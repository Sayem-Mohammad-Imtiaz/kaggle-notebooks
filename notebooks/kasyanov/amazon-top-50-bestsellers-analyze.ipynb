```python
data_path = "../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv"
books_data = pd.read_csv(data_path)
books_data.head()
```

```python
books_data.groupby("Year")["Price"].agg([max, min, np.mean])
```

```python
books_data.groupby("Year")["User Rating"].agg([max, min, np.mean])
```

```python
for_fig1 = books_data.groupby(["Year", "Genre"])["Reviews"].agg([np.sum]).reset_index()
```

```python
for_fig1.head()
```

```python
sns.barplot(x = for_fig1["Year"], y = for_fig1["sum"], hue = for_fig1["Genre"])
```

```python
best_books_by_year = pd.DataFrame(columns = books_data.columns)
for year in range(2009, 2020):
    books_by_year = books_data.loc[books_data["Year"] == year]
    best_books_by_year = pd.concat([best_books_by_year, books_by_year.loc[[books_by_year["User Rating"].idxmax()]]])
best_books_by_year
```

```python
Authors = books_data["Author"].value_counts()
```

```python
Authors_greater_5 = Authors[Authors >= 5]
Authors_greater_5
```

```python
books_data_with_needed_authors = books_data.loc[books_data["Author"].isin(Authors_greater_5.index)]
books_data_with_needed_authors
```

```python
mean_rating = books_data_with_needed_authors.groupby("Author")["User Rating"].mean()
mean_rating = mean_rating.sort_values()
```

```python
plt.figure(figsize = (16, 10))
sns.barplot(x = mean_rating, y = mean_rating.index)
```
