# **Using Normal Decision tree Classifier for building our model**

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
#from adspy_shared_utilities import plot_feature_importances
```

```python
#Including features which have an impact on brand as we have observed from bar plots
x = data[["mpg","cylinders","cubicinches","hp","weightlbs"]].values
```

```python
y=data["brand"].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
```

```python
clf = DecisionTreeClassifier().fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

***We are getting 82% accuracy using decision tree classifier***

# Now we use RANDOM FOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf2 = RandomForestClassifier(n_estimators = 120)
```

```python
clf2.fit(x_train,y_train)
```

```python
clf2.score(x_test,y_test)
```
