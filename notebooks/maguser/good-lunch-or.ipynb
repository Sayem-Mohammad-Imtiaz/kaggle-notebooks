# Analysis

```python
df.groupby('gender')[['math_score', 'reading_score', 'writing_score']].agg(['count', 'median', 'mean'])
```

female has low grade in math, and high score in reading and writing. Why? Pershaps influences race_ethnicity group?

```python
df.groupby(['race_ethnicity', 'gender'])[['math_score', 'reading_score', 'writing_score']].agg(['count', 'median','mean'])
```

Where are differnts?
Group A, B, C, D in math are differents. Group B, D has most different in reading_score
Group A, B, C, D in writing_score are strong different

```python
df.groupby('parental_level_of_education')[['math_score', 'reading_score', 'writing_score']].agg(['count', 'median','mean'])
```

big difference in high school, master's degree in all subjects
there is almost no difference between bachelor's degree and master's degree

```python
df.groupby('lunch')[['math_score', 'reading_score', 'writing_score']].agg(['count', 'median', 'mean'])
```

Interestingly, when choosing a standard lunch, the scores are higher. Proper nutrition? The power of eating habits or parents?

```python
df.groupby(['lunch', 'parental_level_of_education'])[['math_score', 'reading_score', 'writing_score']].agg(['count', 'median', 'mean'])
```

nutrition affects more than parents education

```python
df.groupby('test_preparation_course')[['math_score', 'reading_score', 'writing_score']].agg(['count', 'median', 'mean'])
```

the exam preparation course affected only the grades in reading and writing. Math preparation is weaker.
Maybe it's a good topic - math preparation?

```python
df.boxplot('math_score', by='gender');
```

```python
df.boxplot('writing_score', by='gender');
```

```python
df.groupby(['gender', 'lunch'])[['math_score', 'reading_score', 'writing_score']].agg(['count', 'median', 'mean'])
```

Well-nourished boys are stronger in math.
