```python
Y = data.pop("target")
X = data
```

```python
num_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.3, random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler

trainScaler = MinMaxScaler()
xTrainScaled = trainScaler.fit_transform(X_train[num_cols])
X_train.loc[:, num_cols] = xTrainScaled
X_train.head()
```

```python
testScaler = MinMaxScaler()
xTestScaled = testScaler.fit_transform(X_test[num_cols])
X_test.loc[:, num_cols] = xTestScaled
X_test.head()
```

```python
from sklearn.linear_model import LogisticRegression

model1 = LogisticRegression()
model1_fit = model1.fit(X_train, Y_train)
model1_pred = model1_fit.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model2 = KNeighborsClassifier(n_neighbors=5)
model2_fit = model2.fit(X_train, Y_train)
model2_pred = model2_fit.predict(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB

model3 = GaussianNB()
model3_fit = model3.fit(X_train, Y_train)
model3_pred = model3_fit.predict(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

model4 = DecisionTreeClassifier()
model4_fit = model4.fit(X_train, Y_train)
model4_pred = model4_fit.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

model5 = RandomForestClassifier()
model5_fit = model5.fit(X_train, Y_train)
model5_pred = model5_fit.predict(X_test)
```

```python
from sklearn.svm import LinearSVC

model6 = LinearSVC()
model6_fit = model6.fit(X_train, Y_train)
model6_pred = model6_fit.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, plot_roc_curve, confusion_matrix
```

```python
def evaluation_metrics1(Y_true, pred):
    tn, fp, fn, tp = confusion_matrix(Y_true, pred).ravel()
    acc = accuracy_score(Y_true, pred)
    pre = precision_score(Y_true, pred)
    re = recall_score(Y_true, pred)
    f1 =  f1_score(Y_true, pred)
    return tn, fp, fn, tp, acc, pre, re, f1
```

```python
tn1, fp1, fn1, tp1, acc1, pre1, re1, f1 = evaluation_metrics1(Y_test, model1_pred)
tn2, fp2, fn2, tp2, acc2, pre2, re2, f2 = evaluation_metrics1(Y_test, model2_pred)
tn3, fp3, fn3, tp3, acc3, pre3, re3, f3 = evaluation_metrics1(Y_test, model3_pred)
tn4, fp4, fn4, tp4, acc4, pre4, re4, f4 = evaluation_metrics1(Y_test, model4_pred)
tn5, fp5, fn5, tp5, acc5, pre5, re5, f5 = evaluation_metrics1(Y_test, model5_pred)
tn6, fp6, fn6, tp6, acc6, pre6, re6, f6 = evaluation_metrics1(Y_test, model6_pred)
```

```python
comparision = pd.DataFrame([[tn1, fp1, fn1, tp1, acc1, pre1, re1, f1], [tn2, fp2, fn2, tp2, acc2, pre2, re2, f2], 
                           [tn3, fp3, fn3, tp3, acc3, pre3, re3, f3], [tn4, fp4, fn4, tp4, acc4, pre4, re4, f4], 
                           [tn5, fp5, fn5, tp5, acc5, pre5, re5, f5], [tn6, fp6, fn6, tp6, acc6, pre6, re6, f6]],
                           columns=["True Neg", "False Pos", "False Neg", "True Pos", "accuracy", "precision", "recall", "f1-score"],
                          index=["Logistic Regression", "KNN", "Naive Bayes", "Decision Tree", "Random Forest", "SVM"])
comparision
```
