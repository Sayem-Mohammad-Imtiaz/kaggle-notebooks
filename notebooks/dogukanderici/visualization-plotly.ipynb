# Line Chart

```python
data = times_df.iloc[:100,:]

trace1 = go.Scatter(
    
    x = data.world_rank,
    
    y = data.citations,
    
    mode = "lines",
    
    name = "Citations",
    
    marker = dict(color = "rgba(16,112,2,0.8)"),
    
    text = data.university_name)

trace2 = go.Scatter(
    
    x = data.world_rank,
    
    y = data.teaching,
    
    mode = "lines+markers",
    
    name = "Teaching",
    
    marker = dict(color = "rgba(80, 26, 80, 0.8)"),
    
    text = data.university_name)

data1 = [trace1,trace2]
layout = dict(
    
    title = "Citation and Teaching vs World Rank of Top 100 Universities",
    xaxis = dict(title = "World Ranking",ticklen = 5, zeroline = False)
)
fig = dict(data = data1,layout = layout)
iplot(fig)
```

# Scatter Plot

```python
df2014 = times_df[times_df.year == 2014].iloc[:100,:]
df2015 = times_df[times_df.year == 2015].iloc[:100,:]
df2016 = times_df[times_df.year == 2016].iloc[:100,:]



trace1 = go.Scatter(
    
                        x = df2014.world_rank,
    
                        y = df2014.citations,
    
                        mode = "markers",
    
                        name = "2014",
    
                        marker = dict(color = "rgba(255, 128, 255, 0.8)"),
    
                        text= df2016.university_name
)

trace2 = go.Scatter(
    
                        x = df2015.world_rank,
    
                        y = df2015.citations,
    
                        mode = "markers",
    
                        name = "2015",
    
                        marker = dict(color = "rgba(255, 128, 2, 0.8)"),
    
                        text= df2016.university_name
)

trace3 = go.Scatter(
    
                        x = df2016.world_rank,
    
                        y = df2016.citations,
    
                        mode = "markers",
    
                        name = "2016",
    
                        marker = dict(color = "rgba(0, 255, 200, 0.8)"),
    
                        text= df2016.university_name
)

data2 = [trace1, trace2, trace3]
layout2 = dict(title = "Citations vs world rank of top 100 universities with 2014,2015 and 2016",
             xaxis = dict(title = "World Rank",ticklen = 5, zeroline = False),
             yaxis = dict(title = "Citation",ticklen = 5, zeroline = False))
fig2 = dict(data = data2, layout = layout2)
iplot(fig2)
```

# Bar Chart
* citations and teaching of top 3 universities in 2014 (style 1)

```python
data2014 = times_df[times_df.year == 2014].iloc[:3,:]
trace1 = go.Bar(
    
    x = data2014.university_name,
    y = data2014.citations,
    name = "citations",
    marker = dict(color = 'rgba(255, 174, 255, 0.5)',
                             line=dict(color='rgb(0,0,0)',width=1.5)),
    text = data2014.country
)

trace2 = go.Bar(
    
    x = data2014.university_name,
    y = data2014.teaching,
    name = "teaching",
    marker = dict(color = 'rgba(255, 255, 128, 0.5)',
                              line=dict(color='rgb(0,0,0)',width=1.5)),
    text = data2014.country)

data2014 = [trace1, trace2]
layout = go.Layout(barmode = "group")
fig = go.Figure(data = data2014, layout = layout)
iplot(fig)
```

```python
data2014 = times_df[times_df.year == 2014].iloc[:3,:]
trace1 = {
    "x": data2014.university_name,
    "y": data2014.teaching,
    "name":"teaching",
    "type":"bar"
};

trace2 = {
    "x":data2014.university_name,
    "y":data2014.citations,
    "name":"citatios",
    "type":"bar"
};

data3 = [trace1,trace2];
layout = {
    "xaxis":{"title":"Top 3 Universities in 2014"},
    "barmode":"relative",
    "title":'citations and teaching of top 3 universities in 2014'
};

fig = go.Figure(data = data3, layout = layout)
iplot(fig)
```

# Pie Chart
* Students rate of top 7 universities in 2016

```python
data2016 = times_df[times_df.year == 2016].iloc[:7,:]
data2016
```

```python
data2016.info()
```

```python
# data preparation
data2016 = times_df[times_df.year == 2016].iloc[:7,:]
pie1 = data2016.num_students
pie1_list = [float(each.replace(',', '.')) for each in data2016.num_students]  # str(2,4) => str(2.4) = > float(2.4) = 2.4
labels = data2016.university_name
# figure
fig1 = {
  "data": [
    {
      "values": pie1_list,
      "labels": labels,
      "domain": {"x": [0, .5]},
      "name": "Number Of Students Rates",
      "hoverinfo":"label+percent+name",
      "hole": .3,
      "type": "pie"
    },],
  "layout": {
        "title":"Universities Number of Students rates",
        "annotations": [
            { "font": { "size": 20},
              "showarrow": False,
              "text": "Number of Students",
                "x": 0.20,
                "y": 1
            },
        ]
    }
}
iplot(fig1)
```

# Bubble Plot

```python
data2016 = times_df[times_df.year == 2016].iloc[:20,:]
num_students_size  = [float(each.replace(',', '.')) for each in data2016.num_students]
international_color = [float(each) for each in data2016.international]
data = [
    {
        'y': data2016.teaching,
        'x': data2016.world_rank,
        'mode': 'markers',
        'marker': {
            'color': international_color,
            'size': num_students_size,
            'showscale': True
        },
        "text" :  data2016.university_name    
    }
]
iplot(data)
```

# WordCloud Plot

```python
country_2011 = times_df.country[times_df.year == 2011]
plt.subplots(figsize=(8,8))
wordcloud = WordCloud(
                          background_color='white',
                          width=512,
                          height=384
                         ).generate(" ".join(country_2011))
plt.imshow(wordcloud)
plt.axis('off')

plt.show()
```
