```python
import pandas as pd
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv(‘../input/students-performance-in-exams/StudentsPerformance.csv’)
Gender = pd.get_dummies(df['gender'], drop_first= True)
Race_ethnicity = pd.get_dummies(df['race_ethnicity'], drop_first = True)
Parental_level_of_education = pd.get_dummies(df['parental_level_of_education'], drop_first= True)
Lunch = pd.get_dummies(df['lunch'], drop_first= True)
df2 = pd.concat([df, Gender, Race_ethnicity, Parental_level_of_education,Lunch], axis = 1)


df3 = df2.drop(['gender', 'race_ethnicity', 'parental_level_of_education','lunch'], axis = 1)
```

```python
model = tree.DecisionTreeClassifier(max_depth=2, random_state= 0)
xcol = ['math_score', 'reading_score','writing_score','high school', "master's degree", 'some college','some high school', 'standard']
x =df3[xcol]
t = df3['male']
x_train, x_test, y_train, y_test = train_test_split(x,t,test_size= 0.3, random_state= 0)
model.fit(x_train, y_train)
```

```python
print(model.score(x_test,y_test))
```
