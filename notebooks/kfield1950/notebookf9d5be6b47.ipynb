```python
import pandas as pd
df = pd.read_csv("../input/star-dataset/6 class csv.csv")
df
```

```python
df['Star color'].unique()
df['Star color'].value_counts()
```

```python
df.isnull().any(axis=0)
```

```python
df.loc[df['Star color']=='white','Star color']='White'
df.loc[df['Star color']=='Whitish','Star color']='White'
df.loc[df['Star color'].str.contains('Blue'),'Star color']='Blue'
df.loc[df['Star color'].str.contains('Yellow'),'Star color']='Yellow'
df.loc[df['Star color'].str.contains('yellow'),'Star color']='Yellow'
df.loc[df['Star color'].str.contains('Orange'),'Star color']='Yellow'
                                    

df['Star color'].unique()
df['Star color'].value_counts()
```

```python
df2 = pd.get_dummies(df['Star color'])
df2 = pd.concat([df, df2],axis = 1)

df3 = df2.drop('Star color',axis=1)
df3
```

```python
import seaborn as sns
sns.scatterplot(x='Temperature (K)',y='Absolute magnitude(Mv)',hue='Star type',data=df3)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier(max_depth=5,random_state=0)
```

```python
from sklearn.model_selection import train_test_split
xcol = ['Radius(R/Ro)','Absolute magnitude(Mv)','Blue','Red','Yellow','White']
x = df3[xcol]
t = df3['Star type']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,t,test_size=0.3,random_state=0)
model.fit(x_train,y_train)
model.score(X = x_test, y= y_test)
```
