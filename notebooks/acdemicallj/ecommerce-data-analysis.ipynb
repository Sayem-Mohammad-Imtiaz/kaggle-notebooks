# 2. 数据清洗

2.1 删除缺失值 

```python
df.dropna(inplace=True)
```

```python
df.info()
```

 2.2 异常值处理

```python
df.describe()
```

```python
df = df[(df['Quantity']>0) & (df['UnitPrice']>0)]
```

```python
df.describe()
```

2.3 数据类型转换

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
df.info()
```

# 3. 商品角度分析

3.1 商品总消费情况和月消费趋势

```python
df.groupby('StockCode').Quantity.sum().describe()
```

```python
#查看销量排行
df.groupby('StockCode').Quantity.sum().sort_values(ascending=False)[:10]
```

```python
#查看总体消费情况
plt.figure()
df.groupby('StockCode').Quantity.sum().hist()
plt.xlabel('sales')
plt.show()
```

```python
df['month'] = df.InvoiceDate.astype('datetime64[M]')
```

```python
#查看月消费趋势（总体）
plt.figure()
df.groupby('month').Quantity.sum().plot()
#plt.xlabel('商品销售量')
```

3.2 高销量商品是否受价格影响

```python
data_o = df[['StockCode','UnitPrice']]
data_g = df.groupby('StockCode').Quantity.sum().reset_index()
data_m = pd.merge(data_o,data_g,how='inner',on='StockCode')
data_m.drop_duplicates(subset='StockCode',inplace=True)
data_f = data_m.set_index('StockCode')
data_f.head()
```

```python
data_f.corr()
```

```python
#销量和单价散点图
plt.figure()
plt.scatter(x=data_f.Quantity,y=data_f.UnitPrice)
plt.show()
```

# 4.用户角度分析

4.2 用户月消费金额和购买量趋势

```python
#月消费金额
data_mp = df.groupby('month').UnitPrice.sum()

#月购买量
data_mg = df.groupby('month').Quantity.sum()
```

```python
plt.figure()
data_mg.plot()
plt.show()
```

```python
plt.figure()
plt.scatter(x=data_mg,y=data_mp)
plt.show()
```

```python
plt.figure()
plt.hist(data_mg)
plt.show()
```
