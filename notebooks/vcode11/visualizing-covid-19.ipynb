```python
covid_data = pd.read_csv('/kaggle/input/covid19-in-india/covid_19_india.csv')
covid_data.head(10)
```

```python
covid_data['Date'].value_counts()
```

```python
# for state in covid_data['State/UnionTerritory'].unique():
#     print(state)
covid_data = covid_data[covid_data['State/UnionTerritory'] != 'Cases being reassigned to states']
covid_data = covid_data[covid_data['State/UnionTerritory'] != 'Unassigned']
covid_data['Date'] = pd.to_datetime(covid_data['Date'], dayfirst=True)
covid_data['Date'] = covid_data['Date'].dt.date
covid_data['Active'] = covid_data['Confirmed'] - covid_data['Cured']
# seaborn.lineplot(x='Date', y='Confirmed', data=covid_data[covid_data])
datewise = covid_data.groupby('Date')['Confirmed', 'Cured', 'Deaths', 'Active'].sum()
datewise
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(data=datewise)
plt.title('Number of COVID-19 Cases in India over time', fontsize=14)
plt.ylim([100, 205000])
plt.ylabel('Number of Cases', fontsize=14)
plt.xlabel('Months', fontsize=14)
```

```python
# A bar graph
import datetime
stats_today = covid_data[covid_data['Date'] == datetime.date.today()].sort_values(by=['Confirmed'], ascending=False)
plt.figure(figsize=(14,16))
sns.barplot(y=stats_today['State/UnionTerritory'], x= stats_today['Confirmed'], label='Confirmed').set_title('States vs Total Confirmed Cases')
```

```python
stats_today = covid_data[covid_data['Date'] == datetime.date.today()].sort_values(by=['Active'], ascending=False)
plt.figure(figsize=(14,16))
sns.barplot(y=stats_today['State/UnionTerritory'], x= stats_today['Active'], label='Active').set_title('States vs Number of Active Cases')
```

```python
stats_today = covid_data[covid_data['Date'] == datetime.date.today()].sort_values(by=['Deaths'], ascending=False)
plt.figure(figsize=(14,16))
sns.barplot(y=stats_today['State/UnionTerritory'], x= stats_today['Deaths']).set_title('States vs Number of Deaths')
```

```python
stats_today = covid_data[covid_data['Date'] == datetime.date.today()].sort_values(by=['Cured'], ascending=False)
plt.figure(figsize=(14,16))
sns.barplot(y=stats_today['State/UnionTerritory'], x= stats_today['Cured']).set_title('States vs Number of Cured Cases')
```

```python
covid_data[covid_data['Date'] == datetime.date.today()].groupby('State/UnionTerritory')[['Cured', 'Deaths', 'Active']].sum().sort_values(by=['Active'], ascending=[True]).plot.barh(stacked=True, figsize=(14,16))
```

```python
covid_data[covid_data['Date'] == datetime.date.today()].groupby('State/UnionTerritory')[['Confirmed', 'Cured', 'Deaths']].sum().sort_values(by=['Confirmed'], ascending=[False])
```

```python
datewise_data = covid_data.groupby('Date')[['Confirmed', 'Cured','Active', 'Deaths']].sum()
plt.figure(figsize=(14,12))
sns.regplot(x=datewise_data['Confirmed'], y=datewise_data['Cured'], label='Cured')
sns.regplot(x=datewise_data['Confirmed'], y=datewise_data['Deaths'], label='Deaths')
sns.regplot(x=datewise_data['Confirmed'], y=datewise_data['Active'], label='Active')
plt.title('Cured, Deaths, Active cases vs Confirmed Cases', fontsize=15)
plt.xlabel('Confirmed Cases', fontsize=15)
plt.ylabel('Number Of Cases')
plt.legend()
```
