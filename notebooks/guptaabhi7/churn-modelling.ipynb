```python
df = pd.read_csv("/kaggle/input/churn-modelling/Churn_Modelling.csv")
```

```python
df.head()
```

```python
# drop surname
df.drop("Surname", inplace = True, axis = 1)
```

```python
df.info()
```

```python
df['CustomerId'].nunique()
for col in df:
    if(df[col].nunique() > 5000):
        df.drop(col, inplace = True, axis = 1)
```

```python
df.head()
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()

Geography = label.fit_transform(df['Geography'])
Cred_Score = label.fit_transform(df['CreditScore'])

df.drop(['Geography', 'CreditScore'], inplace = True, axis = 1)
```

```python
df['Geography'] = Geography
df['Credit_Score'] = Cred_Score
```

```python
df.head()
```

```python
gender = pd.get_dummies(df['Gender'], drop_first = True)
df['Gender'] = gender
```

```python
df.head()
```

```python
df.info()
```

```python
X = df.drop('Exited', axis=1)
y = df['Exited']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)

lm.score(X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression()

log.fit(X_train, y_train)

Prediction = log.predict(X_test)

log.score(X_test, y_test)
```

```python
from sklearn.svm import SVC
svc = SVC()

svc.fit(X_train, y_train)

Prediction_SVM = svc.predict(X_test)

svc.score(X_test, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 1)

knn.fit(X_train, y_train)

Prediction_KNC = knn.predict(X_test)

knn.score(X_test, y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
gaussian = GaussianNB()

gaussian.fit(X_train, y_train)

Prediction_G = gaussian.predict(X_test)

gaussian.score(X_test, y_test)
```

```python
## best for this data

from sklearn.ensemble import GradientBoostingClassifier
gradient_boost = GradientBoostingClassifier(n_estimators=100)
gradient_boost.fit(X_train, y_train)

gradient_boost.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, y_train)

random_forest.score(X_train, y_train)
random_forest.score(X_test, y_test)
```
