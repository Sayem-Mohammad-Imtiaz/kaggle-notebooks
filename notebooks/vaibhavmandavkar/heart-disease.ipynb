```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("../input/heart-disease-uci/heart.csv")
data.head()
```

```python
data['target'].unique()
```

```python
data.info()
```

```python
X = data.iloc[:,:-1].values
y = data["target"]
import sklearn.model_selection as model_selection
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.20, random_state=27)
```

```python

from sklearn.naive_bayes import GaussianNB
naive_byes = GaussianNB()
naive_byes.fit(X_train, y_train)
naive_byes_prediction = naive_byes.predict(X_test)

print(classification_report(naive_byes_prediction, y_test))
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=20, random_state=0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
print(classification_report(y_pred.round(),y_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
classifier = AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1),
    n_estimators=200
)
classifier.fit(X_train, y_train)
predictions = classifier.predict(X_test)
print(classification_report(predictions, y_test))
```
