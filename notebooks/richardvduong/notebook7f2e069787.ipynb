```python
# read the dataset into a dataframe called data
data = pd.read_csv("/kaggle/input/ramen-ratings/ramen-ratings.csv")
```

```python
# prints all the rows in the dataframe
for row_name in data.index:
    print(row_name)
    
# prints all the columns in the dataframe
for col_name in data.columns:
    print(col_name)
```

```python
# how to print all values of a certain column
# the datatype is a pandas series
print(data['Country'])
```

```python
# how to print all values within a specific row
# the datatype is a pandas series
print(data.loc[10])
```

```python
# to create a set of the different items in a column (no duplicates)
countries = set()
for row_num in data.index:
    countries.add(data['Country'][row_num])

print(countries)
```

```python
# to print and display the dataframe with only country and stars
for row_num in data.index:
    print(row_num, " ", data['Country'][row_num], " ", data['Stars'][row_num])
```

```python
# to store all the stars for japan and USA in a list
japan_stars = []
usa_stars = []

for row_num in data.index:
    
    if data['Country'][row_num] == "Japan":
        japan_stars.append(data['Stars'][row_num])
        
    elif data['Country'][row_num] == "USA":
        usa_stars.append(data['Stars'][row_num])

# prints the data
print("Japan Stars")
for star in japan_stars:
    print(star)

print("USA Stars")
for star in usa_stars:
    print(star)
```
