```python
df = df.drop('location',axis=1)
df = df.drop('society',axis=1)
df = df.drop('area_type',axis=1)
df = df.drop('availability',axis=1)
df = df.drop('balcony',axis=1)
```

```python
df = df.dropna()
```

```python
df = df.drop_duplicates(keep='first')
```

```python
df['size'] = df['size'].apply(lambda x:x.split()[0])
df['total_sqft'] = df['total_sqft'].apply(lambda x:x.split()[0])
```

```python
df.replace('15Acres',15*43860,inplace=True)
df.replace('5.31Acres',5.31*43860,inplace=True)
df.replace('30Acres',30*43860,inplace=True)
df.replace('2.09Acres',2.09*43860,inplace=True)
df.replace('24Guntha',24*1089,inplace=True)
df.replace('1500Cents',1500*435.6,inplace=True)
df.replace('2Acres',2*43860,inplace=True)
df.replace('1Grounds',1*2400,inplace=True)
df.replace('1.25Acres',1.25*43860,inplace=True)
df.replace('38Guntha',38*1089,inplace=True)
df.replace('1.26Acres',1.26*43860,inplace=True)
df.replace('6Acres',6*43860,inplace=True)
df.replace('3Cents',3*435.6,inplace=True)
df.replace('4125Perch',4125*272.25,inplace=True)
```
