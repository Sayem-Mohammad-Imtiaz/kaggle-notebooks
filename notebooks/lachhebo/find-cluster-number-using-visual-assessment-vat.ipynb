# This kernel aims to show a short example of the **vat algorithm** (visual assessment of tendency)

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from pyclustertend import vat, ivat
from sklearn.preprocessing import normalize

df = pd.read_csv("../input/world-happiness-report-2019.csv")
```

```python
df[df.isnull().any(axis=1)]
```

```python
pd.qcut(df.Corruption, 4)
```

```python
df.Corruption = df.Corruption.fillna(111)
```

```python
dft = df.fillna(-1)
```

```python
X = dft[dft.columns[~dft.columns.isin(["Country (region)"])]]
```

```python
X = normalize(X)
```

```python
vat(X)
```

the visualisation is blurred but we can guess there is stack of observation which seems to form one cluster, we can use a another form of vat which is a refinement of the original algorithm to obtain clearer and more informative visualisation.

```python
ivat(X)
```

We can conclude that nations follow a normal distributions when speeking about happyness, there is a cluster in which most nations are.
