I'm really interested about NBA so let's explore the players who played NBA

```python
nba=data[data['League']=='NBA']
```

```python
# No. of teams in NBA
print('The teams of NBA are :',nba.Team.unique())
```

10 best teams of NBA that scored highest points during 2018-2019 season ....

```python
nba_2018=nba[nba['Season']=='2018 - 2019']
```

```python
score={}
for i in nba_2018.Team.unique():
    score[i]=sum(nba_2018[nba_2018['Team']==i]['PTS'].values)
```

```python
score=sorted(score.items(), key=lambda x: x[1], reverse=True)
```

Let's show the first 10 teams in NBA that scored the highest in the 2018-19 season

```python
score=np.array(score)
```

```python
t=score[:10,0]
s=score[:10,1]
```

```python
plt.bar(t,s)
```

From this we can see that PHI(Philadelphia 76ers) scored the highest points in the 2018-19 series

I wanted to know if there was any school that should be known to make Basketball Players (NBA)

```python
school={}
for i in nba.high_school.unique():
    school[i]=len(nba[nba['high_school']==i]['Player'].unique())
```

```python
school=sorted(school.items(), key=lambda x: x[1], reverse=True)
```

```python
school=np.array(school)
```

```python
sn=school[:10,0]
snm=school[:10,1]
plt.bar(sn,snm)
plt.xticks(rotation=90)
plt.show()
```
