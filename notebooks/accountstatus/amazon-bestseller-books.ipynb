## Preprocessing the data description by removing the stopwords and punctuation ....

```python
stopset = set(stopwords.words('english'))
# Function to remove the puctuation marks from my string
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~1234567890'''
def remove_punctuation(my_str):
    no_punct = ""
    for char in my_str:
       if char not in punctuations:
           no_punct = no_punct + char

    # return the unpunctuated string
    return no_punct
```

```python
def find_words(number=10):
    d=data['Description'].values
    dic={}
    sentences=[]
    for i in d:
        i=remove_punctuation(i)
        tokens=word_tokenize(str(i))
        tokens = [w for w in tokens if not w in stopset]
        for j in tokens:
            if j not in dic:
                dic[j]=1
            else:
                dic[j]+=1
    dic={k: v for k, v in sorted(dic.items(), key=lambda item: item[1],reverse=True)}
    return dict(itertools.islice(dic.items(),18, number+18))
```

```python
find_words(12)
```

```python
t = find_words(16).keys()
text=''
for i in t:
    text+=i+' '
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
