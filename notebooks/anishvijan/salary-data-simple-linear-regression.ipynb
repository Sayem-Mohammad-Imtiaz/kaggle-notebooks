```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# loading the dataset
df = pd.read_csv('../input/salary-data-simple-linear-regression/Salary_Data.csv')
df.head()
```

```python
plt.subplot(1,3,1)
plt.hist(df['YearsExperience'])
plt.xlabel('Years of Experience')
plt.ylabel('count')
plt.title('YoE Plot')

plt.subplot(1,3,2)
plt.boxplot(df['YearsExperience'])
plt.xlabel('Years of Experience')
plt.ylabel('count')
plt.title('YoE Plot')

plt.subplot(1,3,3)
plt.violinplot(df['YearsExperience'])
plt.xlabel('Years of Experience')
plt.ylabel('count')
plt.title('YoE Plot')
```

```python
plt.subplot(1,3,1)
plt.hist(df['Salary'], color='red')
plt.xlabel('Salary')
plt.ylabel('count')
plt.title('Salary Plot')

plt.subplot(1,3,2)
plt.boxplot(df['Salary'])
plt.xlabel('Salary')
plt.ylabel('count')
plt.title('Salary Plot')

plt.subplot(1,3,3)
plt.violinplot(df['Salary'])
plt.xlabel('Salary')
plt.ylabel('count')
plt.title('Salary Plot')
```

```python
plt.scatter(df['YearsExperience'],df['Salary'])
plt.xlabel('YoE')
plt.ylabel('Salary')
plt.title('Scatter Plot')
```

```python
# From above we can see the graph is linear therefore it will be good with linear regression
```

```python
x = df.drop(['Salary'],axis = 1)
x
```

```python
y = df['Salary']
y
```

```python
# train test split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
# training the model
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

model.coef_

```python
model.intercept_
```

```python
# calculating the metrics related to model
```

```python
model.score(X_test,y_test)
```

```python
mean_squared_error(y_test,model.predict(X_test))
```

```python
# Here the model  comes to be 94.4% accurate which is good by just applying simple linear regression.
```

```python
# plottin the graph with prediction line
```

```python
y_pred = X_test.dot(model.coef_)+model.intercept_
```

```python
plt.scatter(x,y)
plt.plot(X_test,y_pred,color = 'red')
plt.xlabel('YoE')
plt.ylabel('Salary')
plt.title('Scatter Plot')
plt.legend(['prediction line','slary'])
```
