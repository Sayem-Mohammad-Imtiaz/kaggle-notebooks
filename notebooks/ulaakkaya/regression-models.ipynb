## Linear Regression
 In Linear Regression model we use a simple line equation which is y=b0+b1*x
* y= y axis of the graph
* x= x axis of the graph
* b0= interception point of y when x=0
* b1= Coefficent(slope of the graph)

In this project I chose to use heart csv file which contains datas such as age, sex, cholesterol rates etc. First, we start with choosing two variables for linear regression from table so I chose age for y axis and cholesterol for x axis.

```python
# taking data frame from the heart.csv file

df=pd.read_csv(r"../input/heart-attack-analysis-prediction-dataset/heart.csv")

# there are a few ways to take values from the table
#I'll show two ways

#here dataset is simplified
#this code takes "age" column and find avarage of the cholesterol and other values that belongs same age
data=df.groupby('age', as_index=False).mean()

#first way
x=data.age.values.reshape(-1,1) #takes all the rows from age column
y=data.chol.values.reshape(-1,1) #takes all the rows from cholesterol column


#second way
x2=data.iloc[:,[0]].values #takes all all rows from first column
y2=data.iloc[:,[4]].values #takes all rows from 4th column

```

```python
# visualization part
plt.xlabel("Age")# naming x axis
plt.ylabel("Cholesterol")# naimng y axis
plt.scatter(x,y)
plt.show()
```

There are two ways to calculate the prediction values: 

    1- we can use the line equation formula 
    2- we can use the linear regression libraries that python provide us.

```python
#% First way

#importing linear regression model from sklearn library
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(x,y)

b0=lr.intercept_
b1=lr.coef_

y_head=b0+x*b1
y_pred=b0+55*b1
print("Cholestrol level prediction at age 55: ",y_pred)

# visualization part
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.scatter(x,y)
plt.plot(x,y_head,color="red")
plt.show()
```

```python
#%% Second way

y_head2=lr.predict(x)
y_pred2=lr.predict([[55]])
print("Cholestrol level prediction at age 55: ",y_pred2)

# visualization part
plt.xlabel("Age")
plt.ylabel("Cholesterol")
plt.scatter(x,y)
plt.plot(x,y_head2,color="red")
plt.show()
```

```python
#%% Finding r square value!!
# imports metrics r square error from sklearn library
from sklearn.metrics import r2_score

r2=r2_score(y,y_head)# calculating r_square error

print(r2)
```

Here we finished linear regression part and as we can see linear regression is not too successful beacuse R square value is realy low. If we want a successful model R square value should be closer to 1.

## Multiple Linear Regression
In multiple linear regression model we apply similar rules with linear regression but this time with choosing two or more column from our data frame as variable x. Formula can be represented as: 

y=b0+b1*x1+b2*x2+b3*x3+....
    
  * b0= interception point of y when x=0
  * b1,b2,b3...= Coefficent(slope of the graph)


```python
x=data.iloc[:,[0,3]].values# takes age column and blood presure column
y=data.iloc[:,[4]].values# takes cholesterol level in mg/dl column

#print(data)
multiple_lr=LinearRegression()
multiple_lr.fit(x,y)
#prediction
y_head=multiple_lr.predict(x)
predict_chol=multiple_lr.predict([[55.7,120]])
print("cholesterol level at age 55 with blood pressure: ",predict_chol)
```

```python
#r square error
r2_s=r2_score(y,y_head)
print("r^2: ",r2_s)
```
