```python
path = '/kaggle/input/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv'


df = pd.read_csv(path)

df.head(5)
```

```python
df.isna().sum()
```

```python
df['Churn'] = df['Churn'].map({'No':0,'Yes':1})
```

```python
df.info()
```

```python
for column in df.columns:
    print('The {} has {} unique values '.format(column,df[column].nunique()))
```

```python
# We see that only 3 columns have more than 5 unique values which are Tenure, MonthlyCharges, TotalCharges. 

#The Tenure columns should be converted to categorical column 
```

```python
def func(x):
    if x < 13:
        return "1"
    elif 12 < x < 25:
        return "2"
    elif 24 < x < 37:
        return "3"
    elif 36 < x < 49:
        return "4"
    elif 48 < x < 61:
        return "5"
    elif 60 < x :
        return "6"
    else:
        return 'N/A'
```

```python
df['tenure']  = df['tenure'].apply(func)
```

```python
df.head()
```

```python
customer_id = ['customerID']
numerical_col = ['MonthlyCharges','TotalCharges']

categorical_col = [ i for i in df.columns if i not in numerical_col and i not in customer_id]
```

```python
df.drop('customerID', inplace = True,axis = 1)

df['TotalCharges'] = df['TotalCharges'].apply(lambda x : 0 if x == ' ' else x)

df['TotalCharges']  = pd.to_numeric(df['TotalCharges'])
```

```python
# EDA
```

```python
for col in categorical_col:
     # get feature
    var = df[col]
    # count number of categorical variable(value/sample)
    varValue = var.value_counts()
    
    # visualize
    plt.figure(figsize = (9,3))
    plt.bar(varValue.index, varValue)
    plt.xticks(varValue.index, varValue.index.values)
    plt.ylabel("Frequency")
    plt.title(col)
    plt.show()
    
```

```python
for col in numerical_col:
    sns.histplot(df[col])
    plt.show()
```

```python
plt.figure(figsize=(50,50))
j = 0
sns.set_theme(style="whitegrid")
for i in categorical_col:
    
    plt.subplot(5,4,j+1)
    sns.histplot(df, x="Churn",  hue=df[i], multiple="stack", palette="light:m_r", edgecolor=".3", linewidth=.5)
    plt.title(f'Distribution of {i}',color = 'black',fontsize = 25)
    j += 1
```
