```python
data = pd.read_csv("/kaggle/input/test-file/tested.csv")
data.head()
```

```python
data.tail()
```

```python
data.drop(['PassengerId', 'Name', 'Ticket', 'Fare'], axis = 1, inplace = True)
```

```python
sns.heatmap(data.isnull(), yticklabels = False)
```

Since Cabin has a lot of Nans, we''ll kick it. Age ko modify.

```python
data.drop("Cabin", axis = 1, inplace = True)
```

```python
sns.countplot(x = "Survived",hue = "Sex" ,data =data)
```

Basically, in this dataset, no females died and no males survived.

```python
sns.boxplot(x = "Pclass", y = "Age", data = data)
```

```python
def age_modify(col):
    Age = col[0] 
    Pclass = col[1]
    
    if pd.isnull(Age):
        if Pclass == 1:
            return 42
        
        elif Pclass == 2:
            return 27

        elif Pclass == 3:
            return 25

    else:
        return Age
```

```python
data["Age"] = data[["Age", "Pclass"]].apply(age_modify, axis = 1)
```

```python
sns.heatmap(data.isnull())
```

```python
data["Sex"] = pd.get_dummies(data["Sex"], drop_first = True)
data.head()
```

```python
data["Embarked"] = pd.get_dummies(data["Embarked"], drop_first = True)
data.head()
```

```python
y = data["Survived"]
data.drop("Survived", inplace = True, axis =1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X, test_X, train_y,test_y = train_test_split(data, y, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(train_X, train_y)


```

```python
preds = model.predict(test_X)
```

```python
from sklearn.metrics import confusion_matrix
confu = confusion_matrix(preds, test_y)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(preds, test_y)
print(confu)
print(accuracy)
```
