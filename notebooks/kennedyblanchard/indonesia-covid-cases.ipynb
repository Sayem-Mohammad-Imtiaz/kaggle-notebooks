```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

print("Below you will find some visual data showing the confirmed covid cases in Indonesia between March and April 2020.")

indonesiacases_filepath = "../input/indonesia-coronavirus-cases/cases.csv"
indonesiacases_data = pd.read_csv(indonesiacases_filepath)


sns.lineplot(data=indonesiacases_data)
plt.figure(figsize=(14,6))
plt.title("Indonesia Coronavirus Cases in March-April 2020")
sns.lineplot(data=indonesiacases_data)


df = pd.read_csv(indonesiacases_filepath)
new_confirmed = list(df['new_confirmed'].fillna(0)).copy()
print(new_confirmed)
date = list(df['date'].fillna(0)).copy()
print(date)


plt.figure(figsize=(10,6))
plt.title("Indonesian Coronavirus Cases in March-April 2020")
indonesiacases_data_index = indonesiacases_data.set_index("date")
indonesiacases_data_index = indonesiacases_data.set_index("new_confirmed")
sns.barplot(x=date, y=new_confirmed)
plt.ylabel("number of cases")


plt.figure(figsize=(10,6))
plt.title("Indonesian Coronavirus Cases in March-April 2020")
sns.distplot(a=indonesiacases_data['new_confirmed'], label="new_confirmed", kde=False)
plt.legend()


plt.figure(figsize=(10,6))
sns.kdeplot(data=indonesiacases_data['new_confirmed'], label="new_confirmed", shade=True)
plt.title("Indonesian Coronavirus Cases in March-April 2020")

print (indonesiacases_data)
```
