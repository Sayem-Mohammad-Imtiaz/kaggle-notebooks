**Inferences:**
* The eldest person encountered/shot was 91 years old male and the youngest were two 6 year old males.
* We can see that the police encounters dominantly involved males which is 96% of total.
* By observing the column 'armed',we can infer that the police had no choice but to shoot the people who were armed,60% of whom carried a gun and 14%were armed with a knife.
* The most targeted race seems to be W followed by B and H.Could we say that there was a racial discrimination?? Further analysis is required.
* Very few(270) of the encountered(5416) were shot and also tasered.This shows that police do not tend to the torture that easily.
* From the countplot of flee column with hue as Armed, it can be observed that,even though in very rare case, unarmed and non-fleeing people were also shot/encountered by the police,the reason for which is still unexplained and needs further analysis.
* California state tops the list with maximum (accounts to 14% of total) number of encounters till date and 10% of the shootings of California happened in Los Angeles city.
* From the countplot of signs_of mental_illness,we can observe no relation between the signs_of mental illness and manner of death and also most of the encountered(75% approx) did not exhibit any signs of mental illness.
* Approximately 60% of the people shot/encountered posed threat one way or the other,we can observe that even the unarmed tried to attack the police.This explains the reason for encounters of a part of the unarmed people who were not fleeing.
* From the timeseries plot,we can observe that there is no decline in the number of deaths/encounters from 2015.Does this mean that the crime rate has been consistent over the years which would imply the failure of protection of law and order in the country by police or is this depicting the ruthless killing of people (approx 1000 per year)by police over the years??To come to a proper conclusion,we need further analysis with other factors leading to encounters also included in the data.
