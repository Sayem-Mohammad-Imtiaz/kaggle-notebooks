## Making New Trump Insult Tweets Using Markov Chains 

In this notebook I am going to use Markov Chains to generate new Trump Insult Tweets

```python
# load in the dataset 
import pandas as pd 

df = pd.read_csv('../input/all-trumps-twitter-insults-20152021/trump_insult_tweets_2014_to_2021.csv')
df.head()
```

I will be using the great [markovify](https://github.com/jsvine/markovify) package - makes using markov chains really easy! 

```python
import markovify 

# combine all of the tweets 
m = markovify.NewlineText('\n'.join(df['tweet']))

# lets generate some text 
m.make_short_sentence(max_chars = 280)
```

The above message was generated by the Markov Chain!  

Next, I am going to generate 10 new tweets using the Markov Chain

```python
for i in range(10):
    tweet = m.make_short_sentence(max_chars = 280)
    print(tweet)
    print('\n')
```

Now we have 10 new tweets generated using the Markov Chain and some of them sound pretty realistic
