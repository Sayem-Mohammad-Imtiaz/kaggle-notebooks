```python
fig = px.scatter(df, x="Total Cases", y="State/UTs" ) #Plotting scatter plots
fig.show()
fig = px.scatter(df, x="Active Ratio (%)", y="State/UTs" )
fig.show()
fig = px.scatter(df, x="Discharge Ratio (%)", y="State/UTs" )
fig.show()
fig = px.scatter(df, x="Deaths", y="State/UTs" )
fig.show()
```

```python
fig = px.line(df, x="State/UTs", y="Total Cases", hover_name="State/UTs", #Plotting a line plot with hover functionality
        line_shape="spline", render_mode="svg")
fig.show()
```

```python
map_cases = px.choropleth( #Plotting active cases via geojson map
    df,
    geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
    featureidkey='properties.ST_NM',
    locations='State/UTs',
    color='Active',
    color_continuous_scale='Reds'
)

map_cases.update_geos(fitbounds="locations")

map_cases.show()
```
