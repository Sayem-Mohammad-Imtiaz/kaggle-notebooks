# Seaborn

```python
sns.set(style = 'darkgrid') #Set background theme
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', style='time') #Hue and style are parameter for understanding the pattern
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind = 'line',sort=True)
```

```python
fmri = sns.load_dataset('fmri')
fmri
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind = 'line',sort=True)
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind = 'line',sort=True, ci='sd') #sd stands for standard deviation
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind = 'line',sort=True, estimator = None)
```

```python
sns.relplot(x='timepoint', y='signal', data=fmri, kind = 'line',sort=True, hue = 'event')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue = 'smoker', col = 'time') #Seperates column with the parameter time
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', data=iris)
```

# Catagorical Data plotting

```python
tips.head()
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
sns.catplot(x='day',y='tip',data=tips, kind='box', hue = 'sex')
```

```python
diamond = sns.load_dataset('diamonds')
diamond
```

```python
sns.catplot(x='color',y='price',data=diamond, kind='boxen') #Boxen is a fancy box plot
```

```python
sns.catplot(x='tip',y='day',data=tips, kind='violin',hue='sex',split=True)
```

# 2 Graphs merged using ax(axis)

```python
a = sns.catplot(x='day',y='tip',data=tips, kind='violin',inner=None)
sns.swarmplot(x='day',y='tip',data=tips, color = 'k', size = 3, ax = a.ax)
```
