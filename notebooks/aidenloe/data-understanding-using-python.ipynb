## Specific Correlations

 Let's investigate the significance (p-value) on **Spending on healthy eating** and **Spending on looks**. 

Here we use the `pearsonr` function from scipy library.

The `help(pearsonr)` tells us that p-value are reasonable for datasets > 500. 

In this case, our dataset is approximately twice that (excluding NAs). So we can use p-value as a gauge.  

The first value is the correlation value and the second is the p-value results. We can see that it is very similar to what found using the pandas `corr` function. 


```python

from scipy.stats.stats import pearsonr

# Check if NA exists
smaller.isnull() # Returns the boolean values for all columns. True = NA exists
smaller.isnull().any()  #tells you which columns have NAs
smaller.isnull().any().any() # tells you if any of the above are True
#smaller.isnull().values.any() alternatively, this works just as well. If any NA exists. 

sub = smaller.dropna() #remove all NAs (drop all NAs)
print("Association between Spending on healthy eating and on looks.")
pearsonr(sub["Spending on healthy eating"], sub["Spending on looks"])


# Another way to remove NAs (drop by secific columns)
# sub2 = smaller.dropna(subset = ['Spending on healthy eating', 'Spending on looks']) 
# pearsonr(sub2["Spending on healthy eating"], sub2["Spending on looks"]) # Run the correlations 
```
