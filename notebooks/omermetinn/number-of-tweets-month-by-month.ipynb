```python
## imports and definiations ##
import datetime
import time
import os

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

pd.options.mode.chained_assignment = None
dirname = "/kaggle/input/tweets-about-the-top-companies-from-2015-to-2020"
```

```python
## loading data ##
tweet = pd.read_csv(os.path.join(dirname, "Tweet.csv"))
company_tweet = pd.read_csv(os.path.join(dirname, "Company_Tweet.csv"))
company = pd.read_csv(os.path.join(dirname, "Company.csv"))
```

```python
## convert timestamp to dates ##
dates = tweet[['tweet_id', 'post_date']]
dates['post_date'] = pd.to_datetime(dates['post_date'], unit='s')
```

```python
## match dates with tickers ##
merged_df = pd.merge(left=company_tweet, right=dates, how='left', left_on='tweet_id', right_on='tweet_id')
ticker_dates = merged_df[['ticker_symbol', 'post_date']]
```

```python
## count the posted tweet number for each day for each company ##
tweet_counts = ticker_dates.groupby(['ticker_symbol', 
                                      ticker_dates['post_date'].dt.to_period("M")]).size().reset_index(level=[0,1])
tweet_counts['post_date'] = tweet_counts['post_date'].apply(lambda x: pd.to_datetime(str(x)))
```

```python
## plot results ##
for ticker_symbol in company['ticker_symbol']:
    company_count = tweet_counts[tweet_counts['ticker_symbol']==ticker_symbol][['post_date', 0]].set_index('post_date')
    plt.plot(company_count, label=ticker_symbol)
    
plt.legend()
plt.ylabel('# of tweets')
plt.xlabel("time")
plt.show()
```
