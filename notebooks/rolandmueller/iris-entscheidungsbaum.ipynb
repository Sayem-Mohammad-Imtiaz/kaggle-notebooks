```python
iris = pd.read_csv("/kaggle/input/iris-flower-dataset/IRIS.csv")
```

```python
iris
```

```python
X = iris.drop(columns="species")
y = iris['species']
```

```python
type(y.values)
```

```python
list(X.columns)
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

# Entscheidungsbaum spezifieren

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(max_depth=3)
```

# Entscheidungsbaum trainieren

```python
clf.fit(X_train, y_train)
```

# Klassen auf den Testdaten vorhersagen

```python
y_predict = clf.predict(X_test)
```

```python
y_predict
```

```python
y_test
```

# Klassifikationsreport ausgeben

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_predict, digits=2))
```

# Entscheidungsbaum ausgeben

```python
from sklearn.tree import plot_tree
```

```python
feature_names = list(X_train.columns)
feature_names
```

```python
class_names = clf.classes_
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8))
plot_tree(clf, feature_names=feature_names, class_names=clf.classes_, filled=True)
plt.show()
```

# Confusion Matrix ausgeben

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_predict)
cm
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
disp = ConfusionMatrixDisplay(cm, display_labels=class_names)
disp.plot(cmap="Blues")
```

```python
from yellowbrick.model_selection import FeatureImportances
```

```python
fi_viz = FeatureImportances(clf, labels=feature_names)
fi_viz.fit(X_test, y_test)
fi_viz.show()
```
