# Datenvorverarbeitung

* y: Labels (Vektor)
* X: Features (Matrix)

* y_pred = f(X)

```python
y = df['species']
```

```python
class_names = list(y.unique())
```

```python
X = df.drop(columns='species')
X
```

```python
feature_names = list(X.columns)
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=10)
```

# K-NÃ¤chste Nachbarn Verfahren

### KNN initialisieren und trainieren

```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
```

```python
clf.fit(X_train, y_train)
```

### KNN anwenden auf Testdaten

```python
y_pred = clf.predict(X_test)
```

```python
y_pred_proba = clf.predict_proba(X_test)
```

### KNN Evaluieren

Vergleichen: 
* y_test: Wahren Klassen (Labels)
* y_pred: Vorhergesagte Klassen (Labels)

```python
from sklearn import metrics
```

```python
print("Accuracy-Score ist", metrics.accuracy_score(y_test, y_pred))
```

```python
print(metrics.classification_report(y_test, y_pred, digits=3))
```

### Confusion Matrix Visualisiern Variente 1

```python
cm = metrics.confusion_matrix(y_test, y_pred)
print(cm)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
disp = ConfusionMatrixDisplay(cm, display_labels=class_names)
disp.plot(cmap="Blues")
plt.show()
```

### Confusion Matrix Visualisieren Variante 2

```python
metrics.plot_confusion_matrix(clf, X_test, y_test, cmap="Blues")
plt.show()
```

### Validation Curve

```python
from yellowbrick.model_selection import ValidationCurve
```

```python
vc_viz = ValidationCurve(clf, param_name="n_neighbors", 
    param_range=np.arange(1, 80, 1), cv=10, n_jobs=-1)
vc_viz.fit(X, y)
vc_viz.show()
```
