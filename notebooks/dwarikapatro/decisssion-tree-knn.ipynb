```python
X = df.drop('class',axis=1)
y = df['class']
```

```python
col = X.columns
le = LabelEncoder()

for i in col:
    X[i] = le.fit_transform(X[i])
```

```python
X.head()
```

# **KNN classification**

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,
                                                test_size=0.25,
                                                random_state=5)
```

```python
algo = {'uniform':[],'distance':[]}
for i in algo.keys():
    for k in range(1,21,2):
        clf = KNeighborsClassifier(n_neighbors=k,weights=i)
        clf.fit(X_train,y_train)
        y_predict = clf.predict(X_test)
        algo[i].append(accuracy_score(y_test,y_predict))
```

```python
plt.plot(algo['uniform'],c='r',marker='o',label='uniform')
plt.plot(algo['distance'],c='g',marker='o',label='distance')
plt.legend()
plt.show()
```

```python
metri = {'1':[],'2':[]}
for i in metri.keys():
    for j in range(1,21,2):
        clf= KNeighborsClassifier(n_neighbors=j,weights='distance',p=int(i))
        clf.fit(X_train,y_train)
        y_pred=clf.predict(X_test)
        metri[i].append(accuracy_score(y_test,y_pred))
```

```python
plt.plot(metri['1'],c='r',marker='o',label='1')
plt.plot(metri['2'],c='g',marker='o',label='2')
plt.legend()
plt.show()
```

**Here we find best for weight and p**

```python
clf= KNeighborsClassifier(n_neighbors=j,weights='distance',p=int(i))
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print('Test accuracy:-',accuracy_score(y_test,y_pred))
```

# **Decision tree classification**

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
cre = {'gini':[],'entropy':[]}
for i in cre.keys():
    for j in range(1,20):
        dec = DecisionTreeClassifier(criterion=i,max_depth=j)
        dec.fit(X_train,y_train)
        y_pred = dec.predict(X_test)
        cre[i].append(accuracy_score(y_test,y_pred))
```

```python
plt.plot(cre['gini'],marker='o',c='r',label = 'gini')
plt.plot(cre['entropy'],marker='o',c='g',label = 'entropy')
plt.legend()
plt.show()
```

```python
dec = DecisionTreeClassifier(criterion='entropy',max_depth=6)
dec.fit(X_train,y_train)
y_pred = dec.predict(X_test)
```

```python
plt.figure(figsize=[12,10])

_ = plot_tree(dec,filled=True,feature_names=X.columns,node_ids=True)
```

```python
print('Test score: ',dec.score(X_test,y_test))
print('Train score: ',dec.score(X_train,y_train))
```
