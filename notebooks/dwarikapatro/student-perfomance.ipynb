```python
def con(x):
    if x>=40:
        return 'pass'
    else:
        return 'fail'
df['math_pass'] = df['math score'].apply(con)
```

```python
def con(x):
    if x>=40:
        return 'pass'
    else:
        return 'fail'
df['reading_pass'] = df['reading score'].apply(con)
df['writing_pass'] = df['writing score'].apply(con)
```

```python
df['over_all']= df.apply(lambda x: 'fail' if x['math_pass']=='fail' or x['reading_pass']=='fail'
                        or x['writing_pass']=='fail' else 'pass',axis=1)
```

**percentage of marks**

```python
df['percentage']= (df['math score']+df['reading score']+df['writing score'])/3
```

```python
df.head()
```

**grade of student**

```python
def con(x):
    if x>=90:
        return 'A+'
    elif x>=80:
        return 'A'
    elif x>=70:
        return 'B'
    elif x>=60:
        return 'C'
    elif x>=50:
        return 'D'
    elif x>=40:
        return 'E'
    else:
        return 'F'
```

```python
df['grade']= df['percentage'].apply(con)
```

```python
df.head()
```

```python
sns.countplot(x= 'grade',data=df)
```

```python
po=df.grade.value_counts()
```

```python
plt.pie(po,labels=po.keys(),autopct='%.0f%%')
plt.show()
```

```python
sns.countplot(x='parental level of education',hue='grade',data=df)
plt.xticks(rotation=45)
plt.show()
```
