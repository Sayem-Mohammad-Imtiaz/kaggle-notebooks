# Geographical Area - Poverty Rate

```python
ppbpl_df.info()
```

```python
ppbpl_df.head(10)
```

```python
#Eyaletlerimize baktık
geouni=ppbpl_df['Geographic Area'].unique()
geouni
```

```python
#poverty_rate içeriğine bakıyoruz
ppbpl_df.poverty_rate.value_counts()
```

```python
# Verilerimizde gördük ki "-" diye anlamsız bir ifade var bunu 0 ile değiştiriyoruz
ppbpl_df.poverty_rate.replace(['-'],0.0,inplace=True)
ppbpl_df.poverty_rate.value_counts()
```

```python
#poverty_rate featuresimiz str bunu floata çevireceğiz
ppbpl_df.poverty_rate=ppbpl_df.poverty_rate.astype(float)
ppbpl_df.info()
```

```python
#Datamızın sıralanmış bir şekilde bara koymak için
area_list=[]
for i in geouni:
    x=ppbpl_df[ppbpl_df['Geographic Area']==i] #benzersiz olan listemizle eşit mi sorusundan true değerleri alanın bütün değerlerini alarak x e atıyoruz
    poverty_rate=sum(x.poverty_rate)/len(x) #poverty_rate oranımızı buluyoruz
    area_list.append(poverty_rate) #bulduğumuz poverty_rate oranlarını bir listede saklıyoruz

new_data_frame=pd.DataFrame({"area_list":geouni,"poverty_rate":area_list}) #yeni datamızı oluşturduk dictionary mantığı ile
new_data_frame=new_data_frame.sort_values(by="poverty_rate",ascending=False)
new_data_frame
```

```python
#datalarımızı görselleştirme zamanı
plt.figure(figsize=(16,10))
sns.barplot(x=new_data_frame.area_list,y=new_data_frame.poverty_rate)
plt.xticks(rotation=90) # eyaletlerin bulunduğu x eksenindeki yazıları 90 derece çevirdik
plt.title("area-povert rate")
```

# proportion of races in the states

```python
srbc_df.head()
```

```python
srbc_df.value_counts()
```

```python
srbc_df.info() # string şeklindeler bunları float ile değiştireceğim
```

```python
srbc_df.replace(['(X)'],0,inplace=True)
srbc_df.replace(['-'],0,inplace=True)
```

```python
srbc_df[['share_white','share_black','share_native_american','share_asian','share_hispanic']] = srbc_df[['share_white','share_black','share_native_american','share_asian','share_hispanic']].astype(float)
# (x) diye bir değer var bu çevrilmiyor üstte onu düzenleyeceğim
type(srbc_df['share_white'][0]) #değerlerimiz artık float
```

```python
area_list=srbc_df["Geographic area"].unique()
sw=[] # her biri bir ırkın istatistiğini tutacak
sb=[]
sna=[]
sa=[]
sh=[]
for i in area_list:
    x=srbc_df[srbc_df["Geographic area"]==i]
    sw.append(sum(x.share_white)/len(x))
    sb.append(sum(x.share_black)/len(x))
    sna.append(sum(x.share_native_american)/len(x))
    sa.append(sum(x.share_asian)/len(x))    
    sh.append(sum(x.share_hispanic)/len(x))
```

```python
#Görselleştirme kısmı
#bar plot ama yandan
plt.figure(figsize=(12,15))
sns.barplot(x=sw,y=area_list,color="green",alpha=0.4,label="white_people")
sns.barplot(x=sb,y=area_list,color="red",alpha=0.5,label="black_people")
sns.barplot(x=sna,y=area_list,color="yellow",alpha=0.6,label="native_american")
sns.barplot(x=sa,y=area_list,color="purple",alpha=0.5,label="asian_people")
sns.barplot(x=sh,y=area_list,color="blue",alpha=0.5,label="hispanic_people")
plt.legend()
plt.show()
```

# each state poverty and high school graduation rate

```python
#Yukarıda proverty ratemiz olan new_data_frame isimli sıralanmış verimiz var
#ayrıca hs_df isimli sıralanmış liseden mezun olma oranları var

new_data_frame.poverty_rate=new_data_frame.poverty_rate/max(new_data_frame.poverty_rate)
hs_df.hs_rate=hs_df.hs_rate/max(hs_df.hs_rate)
new_frame=pd.concat([new_data_frame,hs_df.hs_rate],axis=1)
#data = pd.concat([sorted_data,sorted_data2['area_highschool_ratio']],axis=1)
new_frame.sort_values("poverty_rate",inplace=True)
```

```python
#point plot ile görselleştirme
plt.figure(figsize=(16,10))
sns.pointplot(x="area_list",y="poverty_rate",data=new_frame,color="purple")
sns.pointplot(x="area_list",y="hs_rate",data=new_frame,color="lime")
plt.text(40,0.6,"poverty_rate",color="purple",fontsize=17,style="italic")
plt.text(40,0.56,"hs_rate",color="lime",fontsize=17,style="italic")
plt.xlabel('States',fontsize = 15,color='blue')
plt.ylabel('Values',fontsize = 15,color='blue')
plt.title('High School Graduate  VS  Poverty Rate',fontsize = 20,color='blue')
```
