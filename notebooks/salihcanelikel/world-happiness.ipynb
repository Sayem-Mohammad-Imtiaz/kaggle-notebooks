### The number of countries in each region

```python
happiness_2021['Regional indicator'].value_counts()
```

```python
happiness_2021['Regional indicator'].value_counts().plot(kind="bar",figsize=(10,6))
plt.xticks(rotation=75)
plt.title('Number of countries in each region')
plt.xlabel('Region')
plt.ylabel('Number of Countries')
plt.show()
```

# Plotting the ladder score relation with the region

```python
happiness_2021['Regional indicator']
```

```python
sns.barplot(y=happiness_2021["Ladder score"],x=happiness_2021["Regional indicator"])
plt.xticks(rotation=90)
plt.show()
```

## Mean of ladder score of each region

```python
happiness_2021_mean = happiness_2021.groupby('Regional indicator')['Ladder score'].mean()
print(happiness_2021_mean)
happiness_2021_mean.plot(kind='bar', figsize=(10,10));
```

## Plotting the relation between the ladder score and each aspect

```python
happiness_2021.plot(x="Ladder score", y="Social support", kind="scatter")
happiness_2021.plot(x="Ladder score", y="Healthy life expectancy", kind="scatter")
happiness_2021.plot(x="Ladder score", y="Freedom to make life choices", kind="scatter");
happiness_2021.plot(x="Ladder score", y="Perceptions of corruption", kind="scatter");
```

## Plotting healthy life expectancy with ladder score scatter plot for each region

```python
plot = sns.jointplot(x = 'Healthy life expectancy', y = 'Ladder score', data = happiness_2021, hue = 'Regional indicator', height =10)
```

## Happiest Country

```python
max_index = happiness_2021['Ladder score'].idxmax()
happiest= happiness_2021.iloc[max_index]['Country name']
happiest
```

## Least Happy Country

```python
min_index = happiness_2021['Ladder score'].idxmin()
least_happy= happiness_2021.iloc[min_index]['Country name']
least_happy
```

## Bivariate Variable Analysis

```python
plt.figure(figsize=(15,10))
sns.heatmap(happiness_2021.corr(), annot=True,linewidth=.15)
```

## Turkey Happiness Rate

```python
wh=happiness_2021.sort_values(by=["Ladder score"])
print(wh[wh["Country name"]=="Turkey"]["Ladder score"])
```
