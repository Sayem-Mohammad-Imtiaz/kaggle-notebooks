```python
df_LABS = pd.read_csv('../input/labs.csv')
df_QUES = pd.read_csv('../input/questionnaire.csv')
df_DEMO = pd.read_csv('../input/demographic.csv')
#df_DIET = pd.read_csv('../input/diet.csv')
df_EXAM = pd.read_csv('../input/examination.csv')
```

Variable Name	Variable Description	Data File Name	Data File Description	Begin Year	EndYear	Component	Use Constraints

URXPREG	 Pregnancy test result	UCPREG_H	Pregnancy Test - Urine	2013	2014	Laboratory	None

https://wwwn.cdc.gov/Nchs/Nhanes/Search/variablelist.aspx?Component=Laboratory&CycleBeginYear=2013

```python
labs_p = df_LABS[[not x for x in df_LABS.URXPREG.isna()]][['SEQN','URXPREG']]
labs_p['URXPREG'].describe()
```

CSQ241	Are you currently pregnant or breast feeding a baby?	CSX_H	Taste & Smell	2013	2014	Examination	None

https://wwwn.cdc.gov/Nchs/Nhanes/Search/variablelist.aspx?Component=Examination&CycleBeginYear=2013

```python
exam_p = df_EXAM[[not x for x in df_EXAM.CSQ241.isna()]][['SEQN','CSQ241']]
exam_p['CSQ241'].describe()
```

RIDEXPRG	Pregnancy status for females between 20 and 44 years of age at the time of MEC exam.	DEMO_H	Demographic Variables and Sample Weights	2013	2014	Demographics	None

https://wwwn.cdc.gov/Nchs/Nhanes/Search/variablelist.aspx?Component=Demographics&CycleBeginYear=2013

```python
demo_p = df_DEMO[[not x for x in df_DEMO.RIDEXPRG.isna()]][['SEQN','RIDEXPRG']]
demo_p['RIDEXPRG'].describe()
```

```python
merged_df = labs_p.merge(exam_p.merge(demo_p,on='SEQN'), on='SEQN')
```
