Berarti karena yang diminta adalah visualisasi grafik jumlah pencarian (kolom "value") per satuan waktu (kolom "week_id"). Eh tapi kayaknya kebanyakan deh jumlah sumbu x-nya kalo pake jumlah week, gimana kalo dibagi per jumlah year aja?

```python
dataset["year"], dataset["week"] = dataset.week_id.str.split('-').str
dataset[["week","year"]] = dataset[["week","year"]].astype(int)
```

```python
dataset.head()
```

```python
dataset.year.unique()
```

Nah setelah kolom "week_id" dibagi menjadi "week" dan "year", kita pakai yang "year" aja ya! Jumlahnya cuma 13 data soalnya, jadi lebih mudah bagi yang melihat visualisasi datanya.

```python
select = dataset[dataset.id=="orange"]
```

```python
select.head()
```

Karena hasil pencarian "value" pada dasarnya masih terbagi menjadi per minggu, maka data tiap minggu tersebut ditotal per tahunnya. Masukkan ke variabel baru yaitu "select2" yang kini hanya berisi kolom "year" untuk sumbu x dan kolom "value" untuk sumbu y.

```python
sum(select[select.year==2004].value)
```

```python
select2  = select.groupby("year").sum()["value"]
```

```python
select2.head()
```

Tinggal diplot menggunakan fungsi plot() dan TADAAAAAAA!!! Visualisasi grafik perubahan hasil pencarian makanan "orange" per tahunnya siap disajikan~

```python
select2.plot()
```
