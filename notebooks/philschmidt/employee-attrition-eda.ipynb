# EDA

In order to get an idea, on what factors an employees attrition depends, lets do some bar plots, where we will plot the factors on the x and the average attrition rate for that factor on the y axis.

```python
# convert to numeric for mean
df.Attrition_numeric = df.Attrition
df.loc[df.Attrition == 'Yes','Attrition_numeric'] = 1
df.loc[df.Attrition == 'No','Attrition_numeric'] = 0

plt.figure(figsize=(12,8))
sns.barplot(x = 'Gender', y = 'Attrition_numeric', data=df)
```
