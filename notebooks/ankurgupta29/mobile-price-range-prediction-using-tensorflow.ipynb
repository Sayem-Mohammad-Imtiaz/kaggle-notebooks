```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model

from sklearn.preprocessing import MinMaxScaler




df_test=pd.read_csv('/kaggle/input/mobile-price-classification/test.csv')

X_test=df_test.drop('id',axis=1).values

scaler = MinMaxScaler()
scaler.fit(X_test)
X_test=scaler.transform(X_test)

#print(X_test)
price_range=[]
model = load_model('/kaggle/input/mobile-price/price.h5')
for i in range(len(X_test)):

    X_pred=X_test[i]

    X_pred=X_pred.reshape(1,len(X_pred))
    # print(X_pred)
    y_pred=model.predict(X_pred)
    y_pred=np.argmax(y_pred)
    # print(y_pred)
    price_range.append(y_pred)
    
print(price_range)

new_column=pd.DataFrame({'Price_Range':price_range})

df_test= df_test.merge(new_column,left_index=True,right_index=True)

df_test.to_csv('test1.csv',index=False)
```
