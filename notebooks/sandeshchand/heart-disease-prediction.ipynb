```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X=data.drop('output',axis=1)
data_y=data['output']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

trans=StandardScaler()

data_X=trans.fit(X).transform(X)
data_X.shape,
X_train,X_test,y_train,y_test=train_test_split(data_X,data_y,test_size=.2,random_state=15)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
svc=SVC()
svc.fit(X_train,y_train)
predictions=svc.predict(X_test)
accuracy_svc=round(accuracy_score(y_test,predictions)*100,2)
cf_matrix=confusion_matrix(y_test,predictions)
print(cf_matrix)
print('The accuracy of the SVM model is {} '.format(accuracy_svc))
```

```python
class_dict = {'Normal':0, 'Prone to heart attack':1}
fig,ax=plt.subplots(figsize=(8,8))
g=sns.heatmap(cf_matrix,annot=True,ax=ax)
g.set_yticklabels(g.get_yticklabels(), rotation= 0, fontsize= 12)
g.set_xticklabels(g.get_xticklabels(),rotation=90, fontsize=12)
ax.xaxis.set_ticklabels(class_dict.keys())
ax.yaxis.set_ticklabels(class_dict.keys())
ax.set_xlabel('Predicted lables')
ax.set_ylabel('True lables')
```

```python
from sklearn.naive_bayes import GaussianNB
guassian=GaussianNB()
guassian.fit(X_train,y_train)
predictions=guassian.predict(X_test)
acc_guassian=round(accuracy_score(y_test,predictions)*100,2)
print("The accuracy of NB model is {}".format(acc_guassian))
```

```python
from sklearn.tree import DecisionTreeClassifier
decisionTree=DecisionTreeClassifier()
decisionTree.fit(X_train,y_train)
predicitons=decisionTree.predict(X_test)
acc_DT=round(accuracy_score(y_test,predictions)*100,2)
print("The accuracy of DT model is {}".format(acc_DT))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
predicitons=rf.predict(X_test)
acc_rf=round(accuracy_score(y_test,predictions)*100,2)
print("The accuracy of rf model is {}".format(acc_rf))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbk=GradientBoostingClassifier()
gbk.fit(X_train,y_train)
predicitons=gbk.predict(X_test)
acc_gbk=round(accuracy_score(y_test,predicitons)*100,2)
print("The accuracy of the gbk is {}".format(acc_gbk))
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(X_train,y_train)
predictions=logmodel.predict(X_test)
acc_log=round(accuracy_score(y_test,predictions)*100,2)
print("The accuracy of the log_model is {}".format(acc_log))
```

```python
models=pd.DataFrame({'Model':['Suppor vector Mechanism','Decision Tree','Random Forest','Guassian Naive Bayes',
                              'Gradient Boosting Classifier','Logistic Regression'],
                     'Score':[accuracy_svc,acc_DT,acc_rf,acc_guassian,acc_gbk,acc_log]})

models.sort_values(by='Score',ascending=False)
models
```
