```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
        
movie_plots = pd.read_csv('/kaggle/input/wikipedia-movie-plots/wiki_movie_plots_deduped.csv')
movies = pd.read_csv('../input/rotten-tomatoes-movies-and-critic-reviews-dataset/rotten_tomatoes_movies.csv')


movies['release_year'] = movies['original_release_date'].str.slice(0,4)
movies['release_year'] = movies['release_year'].astype(str)

# Release Year is an int and doesnt like to compare to str
movie_plots['Release Year'] = movie_plots['Release Year'].astype(str)

```

```python
# merge both datasets
merged_df = pd.merge(movies, movie_plots, left_on = ["movie_title", "release_year"], right_on = ["Title", "Release Year"], how = "inner")

# filter by scifi
scifi_df = merged_df[merged_df['genres'].str.contains("(?i)science fiction", na=False)].copy()

# create new col kiss based on if plot has kiss in content
scifi_df['kiss'] = scifi_df['Plot'].apply(lambda x: 'Yes' if 'kiss' in x.lower() else 'No')

# creating a copy to save as file and to use for plotting
scifi_df_filtered = scifi_df[['Title','tomatometer_rating','kiss']].copy()
scifi_df_filtered.to_csv('scifi_df_filtered.csv',index=False)
scifi_df_filtered
```

```python
sns.violinplot(x="kiss", y="tomatometer_rating", data=scifi_df_filtered, inner="quartile")
```

```python
# adding a dummy col cuz seaborn doesn't like x of None
scifi_df_filtered['dummy'] = 1

# plotting with scale to how many entries are there
sns.violinplot(x='dummy',y="tomatometer_rating", hue="kiss",

                    data=scifi_df_filtered, split=True,

                    scale="count", inner="quartile")
```
