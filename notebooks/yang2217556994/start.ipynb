```python
#1.数据集有多少场战役，每场战役战争数
#这个问题又可以提出几个其他小问题
#A.战役总数量
#B.每年多少次战役，多少次战争，
#C.战争数量的年趋势
num_of_battle_name = battles_df['name'].unique()
print u'战役总数量:',len(num_of_battle_name)
#每年多少次战役
num_of_battle_by_year = battles_df.groupby('year')['name'].count()
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
sns.barplot(num_of_battle_by_year.index,num_of_battle_by_year.values)
ax1.set_title(u'每年战役数量')
ax1.set_ylabel(u'数量')
plt.show()
```

```python
#每年多少次战争
num_of_battlesnum_by_year = battles_df.groupby('year')['battle_number'].sum()
fig = plt.figure()
ax2 = fig.add_subplot(1,1,1)
sns.barplot(num_of_battlesnum_by_year.index,num_of_battlesnum_by_year.values)
ax2.set_title(u'每年战争场数')
ax2.set_ylabel(u'数量')
plt.show()
```

```python
#战争数量的年趋势
num_of_battlesnum_by_year.plot()
```

```python
#2.战争攻击方和防御方情况
#进攻方情况
attack_side = battles_df.groupby('attacker_king')['attacker_king'].count().sort_values(ascending = False)
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
x = attack_side.values
labels = attack_side.index
explode = (0,0,0,0)
ax.pie(x,explode=explode,labels=labels,autopct='%1.1f%%',shadow=False,startangle=90) #x为值，explode表示是否有分离，labels是标签，autopct百分比数据标签，startangle开始角度
ax.axis('equal')
ax.set_title(u'进攻各方')
plt.show()
#看的出来joffrey方主动进攻次数最多，同时剧中他们那一方的实力在人类也算最强大的一列
```

```python
#防御方情况
defend_side = battles_df.groupby('defender_king')['defender_king'].count().sort_values(ascending=False)
print defend_side
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
x = defend_side.values
labels = defend_side.index
explode = (0,0,0,0,0,0)
ax.pie(x,explode=explode,labels=labels,autopct='%1.1f%%',shadow=False,startangle=90)
ax.axis('equal')
plt.title(u'防御各方')
plt.show()
#可以看出，相当长一段时间也的确是Robb Stark跟Joffrey的战争，而像Renly和Mance基本上都是第一场战争就失败了
```

```python
#进攻方及其附属
attack_side_follower = battles_df.groupby('attacker_king')['attacker_1','attacker_2','attacker_3','attacker_4'].count()
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
x = np.arange(4)
width = 0.25
ax.bar(x,attack_side_follower.iloc[:,0].values,width,label='attacker-1')#label添加图例
ax.bar(x+width,attack_side_follower.iloc[:,1].values,width,color='r',label='attacker_2')
ax.bar(x+2*width,attack_side_follower.iloc[:,2].values,width,color='green',label='attacker-3')
ax.bar(x+3*width,attack_side_follower.iloc[:,3].values,width,color='C',label='attacker-4')
ax.set_xticks(x+width)
ax.set_xticklabels(attack_side_follower.iloc[:,0].index,rotation='vertical')#rotation让X轴标签倒置
plt.title(u'进攻方及其附属情况')
plt.legend(loc='best')#图例自动放在最好位置
plt.show()
```

```python
#防御方及其附属
defend_side_follower = battles_df.groupby('defender_king')['defender_1','defender_2','defender_3','defender_4'].count()
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
x = np.arange(6)
width = 0.25
ax.bar(x,defend_side_follower.iloc[:,0].values,width,label='defender_1')
ax.bar(x+width,defend_side_follower.iloc[:,1].values,width,color='r',label='defender_2')
ax.bar(x+2*width,defend_side_follower.iloc[:,2].values,width,color='green',label='defender_3')
ax.bar(x+3*width,defend_side_follower.iloc[:,3].values,width,color='C',label='defender_4')
plt.title(u'防御方及其附属情况')
ax.set_xticks(x+width)
ax.set_xticklabels(defend_side_follower.iloc[:,1].index,rotation='vertical')
plt.legend(loc='best')
plt.show()
```

```python
#7.战争主要发生在哪几个地方
happens_region = battles_df.groupby('region')['region'].count().sort_values(ascending=False)
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
rect = ax.bar(np.arange(len(happens_region.index)),happens_region.values)#画柱形图（x为有多少根柱子，y为值）
x = []
for rec in rect:
    x.append(rec.get_x()+rec.get_width()/2)#为了刻度能够准确，所以要得到x的值加上x的宽度除以2 的值
x = tuple(x)
plt.xticks(x,happens_region.index)#这个是plt里的用法，里面的x要一个元组，还有标签
ax.set_xticklabels(happens_region.index,rotation='vertical')
for rec in rect: #画数据标签
    height = rec.get_height()  #得到每根柱子的高度
    plt.text(rec.get_x()+rec.get_width()/2.4,1.01*height,'%s'%int(height)) #用plt.text（x,y,text）方法，x是x的位置，y是高度，最后text是要画的数据标签
plt.title(u'战争主要发生地')
plt.show()
```

```python
#character_deaths_df死去角色的特征，第二个数据集
#1.死亡人数跟盟友的关系，Allegiances
def str_split(df):
    if 'House' in df:
        df = df.split(" ")[1].strip()
        return df
    return df
character_deaths_df['Allegiances'] = character_deaths_df['Allegiances'].apply(str_split)
allegiances = character_deaths_df.groupby('Allegiances')['Name'].count().sort_values(ascending=False)
sns.barplot(allegiances.index[1:6],allegiances.values[1:6])
plt.title(u'死亡人数前5的势力')
#刚好也是剧中主要活跃的几个家族
```

```python
#死亡人数中性别有什么影响Gender
def gender_tan(df):
    if df == 1:
        return 'Male'
    else:
        return 'Female'
#character_deaths_df['Gender'] = character_deaths_df['Gender'].apply(gender_tan)
gender = character_deaths_df.groupby('Gender')['Name'].count().sort_values(ascending=False)
sns.barplot(gender.index,gender.values)
plt.title(u'男女死亡人数')
#还是男性死得多，这也正是那个时代主要的政治经济活动都是男性参与
```

```python
#Nobility看一下身份与死亡人数的关系
nobility = character_deaths_df.groupby('Nobility')['Name'].count().sort_values(ascending=False)
sns.barplot(nobility.index,nobility.values)
plt.title(u'身份与死亡人数')
#看起来是平民死得更多，这也印证了一个观点，战争总是平民最遭殃
```
