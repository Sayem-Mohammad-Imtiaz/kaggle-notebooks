## Word Clouds of Ingredients Used in Indian Foods

```python
import urllib
import requests
from PIL import Image
from wordcloud import WordCloud,ImageColorGenerator

urllib.request.urlretrieve("https://www.ubackground.com/_ph/9/699547809.jpg", "ingredients.jpg")

img=PIL.Image.open("ingredients.jpg")

# combining the image with the dataset
Mask = np.array(img)

# I've used the ImageColorGenerator library from Wordcloud 
# Here I took the color of the image and impose it over our wordcloud
image_colors = ImageColorGenerator(Mask)
img
```

```python
print("Number of Different Ingredients Used in the Variety of Indian Foods:",len(set(ifood2['ingredients'].values)))
```

```python
## Method for creating wordclouds

def display_cloud(data,title,image_colors):
    wc = WordCloud(background_color='white', max_words=500, max_font_size=256,random_state=42,width=Mask.shape[1],height=Mask.shape[0],
                   mask=Mask,color_func=image_colors)
    wc.generate(' '.join(data))
    sns.set_context("talk", font_scale=.9)
    plt.figure(figsize=(20,8))
    plt.imshow(wc,interpolation="bilinear")
    plt.title(title)
    plt.axis('off')
    plt.show()
```

### Word Cloud of Ingredients Used across India

```python
display_cloud(set(ifood2['name'].values),'Food Names\n',image_colors)
display_cloud(set(ifood2['ingredients'].values),'Variety of Ingredients Used in Indian Foods\n',image_colors)
```
