# **Data Discretization and Visualization**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("../input/hotel-booking-demand/hotel_bookings.csv")
df.head(n = 10)
```

```python
df.describe()
```

# **Visualizing data**

# 1.**Pie chart depicting types of meal ordered by customers**

```python
dict_meal = {}
dict_meal = df.value_counts(df['meal'])
meal_x = ['BB', 'HB', 'SC', 'Undefined', 'FB']
meal_y = [92310, 14463, 10650, 1169, 798]
myexplode = [0.2, 0.2, 0.2, 0.8, 0.8]
plt.pie(meal_y, labels = meal_x, radius=1.8, autopct='%0.01f%%', shadow=True, explode=myexplode)
```

# **2.bar graph showing what majority of customers do(ie. either they cancel their reservation or not).**

```python
sns.countplot(data=df, x = 'is_canceled')
plt.show()
```

# **3.bar graph showing inclination of customers towards which type of room**

```python
sns.countplot(data=df, x = 'reserved_room_type')
plt.show()
```

# **4.Scatter diagram for looking at which type of room is assigned and reserved by customers**

```python
df.plot(kind = 'scatter', x = 'reserved_room_type', y = 'assigned_room_type', color = '#88c999')
plt.show()
```

# **5. Histogram showing at what time do customers book rooms (ie. on arrival or reserved).**

```python
df['lead_time'].plot(kind = 'hist')
```
