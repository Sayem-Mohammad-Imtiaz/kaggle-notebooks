```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../input/salary-data-simple-linear-regression/Salary_Data.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#check for null values

df.isnull().sum()
```

```python
plt.boxplot(df.YearsExperience)
plt.show()
```

```python
plt.boxplot(df.Salary)
plt.show()
```

```python
def normalize(x):
    min = x.min()
    max = x.max()
    y = (x-min)/(max-min)
    return y
```

```python
df1 = normalize(df)
df1.describe()
```

```python
plt.boxplot(df1.YearsExperience)
plt.show()
```

```python
plt.boxplot(df1.Salary)
plt.show()
```

```python
sns.pairplot(df1, x_vars='YearsExperience', y_vars='Salary')
```

**To visualize better lets increase the size of the plot**

```python
sns.pairplot(df1, x_vars='YearsExperience', y_vars='Salary', height=5)
```

```python
X = df1.Salary

X.head()
```

```python
y = df1.YearsExperience

y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7,random_state=30)
```

```python
X_train=X_train[:, np.newaxis]
X_test=X_test[:, np.newaxis]
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()

lr.fit(X_train,y_train)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
y_pred = lr.predict(X_test)
```

```python
c = [i for i in range(1,10,1)]
fig=plt.figure()
plt.plot(c,y_test,color='blue', linewidth=2.5, linestyle='-')
plt.plot(c,y_pred,color='red', linewidth=2.5, linestyle='-')
fig.suptitle('Actual v/s Predicted')
#plt.xlabel('YearsExperience')
#plt.ylabel
```

```python
from sklearn.metrics import mean_squared_error, r2_score

mse=mean_squared_error(y_test,y_pred)
r2_square = r2_score(y_test,y_pred)
```

```python
print('Mean Squared Error:', mse)
print('r squarevalue:', r2_square)
```

****ie., the model built is 96% accurate****
