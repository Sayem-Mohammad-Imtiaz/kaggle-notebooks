I created this function to find max values for every each column.

```python
def findmaxvalue(x):
    t = 0
    for value in x:
        if value>t:
            j=0
            t = value
        if t == value:
            j+=1
    return t,j
```

```python
def findcorrvalue(x,maxi,j):
    liste = []
    a=2
    for y in range(1,j+1):
        i = 0
        for value in x.values:
            if a!=i:
               if value == maxi:
                    liste.append(i)
                    break
            i = i + 1
        a=i
    return liste
```

```python
def corrvalue(x,y):
    value = []
    for t in x:
        value.append(y[t])
    return value
```

max setosa values.

```python
max_setosa_sepal_length=findmaxvalue(setosa["sepal_length"])[0]
max_setosa_sepal_width = findmaxvalue(setosa["sepal_width"])[0]
max_setosa_petal_length = findmaxvalue(setosa["petal_length"])[0]
max_setosa_petal_width = findmaxvalue(setosa["petal_width"])[0]
```

```python
count_setosa_sepal_length = findmaxvalue(setosa["sepal_length"])[1]
count_setosa_sepal_width =   findmaxvalue(setosa["sepal_width"])[1]
count_setosa_petal_length = findmaxvalue(setosa["petal_length"])[1]
count_setosa_petal_width = findmaxvalue(setosa["petal_width"])[1]
```

```python
corr_setosa_sepal_length = corrvalue(findcorrvalue(setosa["sepal_length"],max_setosa_sepal_length,count_setosa_sepal_length),setosa["sepal_width"])
corr_setosa_sepal_width = corrvalue(findcorrvalue(setosa["sepal_width"],max_setosa_sepal_width,count_setosa_sepal_width),setosa["sepal_length"])
corr_setosa_petal_length = corrvalue(findcorrvalue(setosa["petal_length"],max_setosa_petal_length,count_setosa_petal_length),setosa["petal_width"])
corr_setosa_petal_width = corrvalue(findcorrvalue(setosa["petal_width"],max_setosa_petal_width,count_setosa_petal_width),setosa["petal_length"])
```

average setosa values.

```python
avg_setosa_sepal_length = findaverage(setosa["sepal_length"])
avg_setosa_sepal_width =  findaverage(setosa["sepal_width"])
avg_setosa_petal_length =  findaverage(setosa["petal_length"])
avg_setosa_petal_width =  findaverage(setosa["petal_width"])
```
