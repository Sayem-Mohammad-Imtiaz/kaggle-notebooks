Beat forecast is one of the most notable utilizations of AI and information science in the Customer Relationship Management (CRM) and Marketing fields. Basically, a churner is a client or client that quits utilizing an organization's items or administrations.

Beat applications are regular in a few areas: 

**Membership business organizations** (think web and telephone utilities suppliers): clients that are well on the way to beat toward the finish of their membership are reached by a call place and offered a markdown. 

**Web based business organizations** (think Amazon and so forth): programmed messages are sent to clients that haven't purchased anything for quite a while, however may react to a limited time special.

```python
#important library
!pip install dabl
import pandas as pd
import pandas_profiling as npp
import dabl
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from xgboost import XGBClassifier
```

```python
#read data
data = pd.read_csv('/kaggle/input/churn-modelling/Churn_Modelling.csv')
df= data.copy()
```

```python
#EDA
profile = npp.ProfileReport(data)
profile
```

```python
#label enconding
label_encoder = preprocessing.LabelEncoder() 
data['Gender']= label_encoder.fit_transform(data['Gender']) 
data['Geography']= label_encoder.fit_transform(data['Geography'])
```

```python
#drop columns
data.drop('Surname',axis=1,inplace=True)
```

```python
#graphs
df_hello = dabl.clean(data, verbose=1)
types = dabl.detect_types(df_hello)
types
```

```python
#exicited
dabl.plot(df_hello, target_col="Exited")
```

```python
ec = dabl.SimpleClassifier(random_state=0).fit(df_hello, target_col="Exited") 
```

```python
#standard scaling
X = data.drop('Exited', axis =1)
y = data['Exited']
X = StandardScaler().fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
log_reg = LogisticRegression(class_weight='balanced', max_iter=1000, solver='lbfgs')
log_reg.fit(X_train, y_train)
```

```python
y_predicted = log_reg.predict(X_test)
y_predicted
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sn
import matplotlib.pyplot as plt
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('actual')
```

```python
X = df.iloc[:, 3:13].values
y = df.iloc[:, 13].values
```

```python
# Encoding categorical data
labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])
labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])
onehotencoder = OneHotEncoder(categorical_features=[1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
classifier = XGBClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
# Making the confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
# Applying K-Fold Cross Validation
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10)
accuracies.mean()
```
