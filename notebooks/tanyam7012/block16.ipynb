Посчитать статистические показатели:
- медиана,
- мода,
- средняя,
- минимум,
- максимум

```python
print("Медиана = ", df.Reviewer_Score.median())
print("Мода:\n", df.Reviewer_Score.mode())
print("Средняя = ", df.Reviewer_Score.mean())
print("Минимум = ", df.Reviewer_Score.min())
print("Максимум = ", df.Reviewer_Score.max())
```

Создать новый столбец, в котором разместить разницу между
максимумом и минимумом.


```python
df['max-min'] = df.Reviewer_Score.max()-df.Reviewer_Score.min()
df
```

Разбить данные на несколько блоков (не менее 3, приблизительно равных
по размеру) по некоторому критерию (критерий определить
самостоятельно).

```python
block1 = df.Reviewer_Score[df.Reviewer_Score < 6]
block2 = df.Reviewer_Score[(df.Reviewer_Score >= 6) & (df.Reviewer_Score < 9)]
block3 = df.Reviewer_Score[df.Reviewer_Score >= 9]

print("block1 =\n",block1)
print("block2 = \n",block2)
print("block3 = \n",block3)
```

Привести круговую диаграмму для проделанного разбиения.

```python
from matplotlib.pyplot import pie, axis, show


values = [block1.count(), block2.count(),block3.count()]
labels = ['<= 6','от 6 до 9','>= 9']
pie(values, labels=labels);
show()
```

Сгруппировать данные (GroupBy) по некоторому признаку и сохранить
результаты в новые таблицы. Для каждой новой таблицы провести
сортировку по сложному ключу, состоящему из нескольких признаков.
Для каждой таблицы своя сортировка.

```python
data_grp = df.groupby('Hotel_Name')
grp1 = data_grp.get_group("Hotel Arena") 
grp2 = data_grp.get_group("K K Hotel George") 
grp3 = data_grp.get_group("Atlantis Hotel Vienna")

grp1.sort_values(by = ['Review_Total_Negative_Word_Counts', 'Reviewer_Score'], ascending = [False, True]).reset_index(drop = True)
grp2.sort_values(by = ['Average_Score', 'Additional_Number_of_Scoring']).reset_index(drop = True) #по умолчанию True
grp3.sort_values(by = ['Total_Number_of_Reviews_Reviewer_Has_Given', 'Reviewer_Score'], ascending = [False, False]).reset_index(drop = True)
```
