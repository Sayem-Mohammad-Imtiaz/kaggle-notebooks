```python
df = pd.read_csv(r"/kaggle/input/the-human-freedom-index/hfi_cc_2020.csv")
dfw=df.head(50)

#media, std, kurtosis , skewness by columns
dfw.mean(axis = 0)
dfw.std(axis= 0)
dfw.kurtosis(axis = 0)
dfw.skew(axis = 0)
```

```python


#selecting only non-categorical data (lol, also year, makes non sense...)
dfnumbers=df.select_dtypes(exclude=['object'])

#normalize
ndf=(dfw["hf_score"]-dfw["hf_score"].mean(axis=0))/dfw["hf_score"].std(axis=0)

#representing normalized data first, testing by "hf_score"
print("Mean:")
print(ndf.mean(axis=0))
print("Symmetric:")
print(ndf.skew(axis=0))
print("Kurtosis:")
print(ndf.kurtosis(axis=0))

sns.displot(ndf,kind="kde")
plt.plot([ndf.mean(axis=0),ndf.mean(axis=0)],[0,1])
plt.plot([ndf.mean(axis=0)-ndf.std(axis=0),ndf.mean(axis=0)-ndf.std(axis=0)],[0,1])
plt.plot([ndf.mean(axis=0)-2*ndf.std(axis=0),ndf.mean(axis=0)-2*ndf.std(axis=0)],[0,1])
plt.plot([ndf.mean(axis=0)+ndf.std(axis=0),ndf.mean(axis=0)+ndf.std(axis=0)],[0,1])
plt.plot([ndf.mean(axis=0)+2*ndf.std(axis=0),ndf.mean(axis=0)+2*ndf.std(axis=0)],[0,1])
```

```python
#Now we are going to check for each param
i=0 #well...better for firsts 5 params

#same steps as before
for column in dfnumbers.columns:
    
    ndf=(dfw[column]-dfw[column].mean(axis=0))/dfw[column].std(axis=0)

    print("Mean:")
    print(ndf.mean(axis=0))
    print("Symmetric:")
    print(ndf.skew(axis=0))
    print("Kurtosis:")
    print(ndf.kurtosis(axis=0))

    sns.displot(ndf,kind="kde")
    plt.plot([ndf.mean(axis=0),ndf.mean(axis=0)],[0,1])
    plt.plot([ndf.mean(axis=0)-ndf.std(axis=0),ndf.mean(axis=0)-ndf.std(axis=0)],[0,1])
    plt.plot([ndf.mean(axis=0)-2*ndf.std(axis=0),ndf.mean(axis=0)-2*ndf.std(axis=0)],[0,1])
    plt.plot([ndf.mean(axis=0)+ndf.std(axis=0),ndf.mean(axis=0)+ndf.std(axis=0)],[0,1])
    plt.plot([ndf.mean(axis=0)+2*ndf.std(axis=0),ndf.mean(axis=0)+2*ndf.std(axis=0)],[0,1])
    plt.show()
    if(i==4):
        break
    i+=1
```
