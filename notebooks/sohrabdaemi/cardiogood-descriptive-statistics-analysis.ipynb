```python
df = pd.read_csv('/kaggle/input/cardiogoodfitness/CardioGoodFitness.csv')

print(df.head(),'\n')
print(df.describe(),'\n')
print(df.info())

```

```python
!pip install seaborn==0.11.0
import seaborn as sns
print(sns.__version__)
```

```python
#Quick overview of the data

sns.set_style('white')
sns.pairplot(df)
plt.show()
```

```python
#Overview of the correlation of the different variables
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
#Which is the most popular model by gender?
sns.countplot(x='Product',data=df,dodge=True, hue='Gender')
plt.show()
```

```python
#How do income and age affect the decision of which model is bought? We can infer that TM798 is the more expensive model
sns.scatterplot(x='Age', y='Income',data=df, hue = 'Product')
plt.show()
```

```python
#How about marital status? Seems that people in relationships are overall more likely to purchase the treadmill
sns.histplot(x='Product',data=df,hue='MaritalStatus',alpha=0.8, element='bars',stat='density',multiple='dodge')
plt.ylabel('Normalized count')
plt.show()
```

```python
#Boxplots offer a concise summary of certain key aspects of the data as well eg: TM195 and TM498 are bought by people in approximately the same age ranges, but different income brackets
#TM498 is mainly purchased by higher income individuals with a high fitness level
x = 'Product'
y = list(['Age','Income','Miles'])

fig, ax = plt.subplots(3,1,figsize=[7,15])
ax = ax.flatten()
for i in range(len(ax)):
    sns.boxplot(x=x,y=y[i],data=df,ax=ax[i])

plt.show()

```

```python
#Some information on how many miles are run per week or per session: TM498 is used for longer sessions
df['Miles per session'] = df['Miles']/df['Usage']
sns.histplot(x='Miles per session', data=df)
plt.axvline(np.mean(df['Miles per session']),color='r',linestyle='--')
plt.xlabel('Miles per session')
plt.show()

sns.histplot(x='Miles',data=df,hue='Product',multiple='dodge')
plt.axvline(np.mean(df['Miles']),color='r',linestyle='--')
plt.xlabel('Miles per week')
plt.show()

```

```python
#Education level is directly correlated with income as highlighted in the pairplot and correlation heatmap above, so highly educated indviduals are more likely to purchase the more expensive model 
fig, ax = plt.subplots(figsize=[10,5])
sns.histplot(x='Product',data=df,hue='Fitness',alpha=0.5, element='bars',stat='density',multiple='dodge')
plt.ylabel('Normalized count')
plt.show()
sns.stripplot(x='Product',y='Education',data=df)
plt.show()
sns.stripplot(x='Product',y='Income',data=df)
plt.show()
```
