```python
def testNeighborsNumber(n_neighbors_arr, train_X, val_X, train_y, val_y):   
    for n in n_neighbors_arr:
        knn = KNeighborsClassifier(n)
        knn.fit(train_X, train_y)
        score = knn.score(val_X, val_y)
        print("Test set score {} : {:.5f}".format(n, score))
        
```

```python

#Load input data
stroke_dataset_path = '../input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv'
stroke_df = pd.read_csv(stroke_dataset_path)

#Clean Empty Data and Convert gender
filtered_stroke_df = stroke_df.replace(['Male', 'Female', 'Other'],[0,1,np.nan]).dropna(axis=0)

#Define the target
y = filtered_stroke_df.stroke

#Define the features
features = ['age','gender', 'hypertension', 'heart_disease','bmi', 'avg_glucose_level']
X = filtered_stroke_df[features]
X = scale(X)

#Split data on train and validation data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=10)

testNeighborsNumber([5,6,7,8,9,10], train_X, val_X, train_y, val_y)
```

```python
from sklearn.neural_network import MLPClassifier

#Load input data
stroke_dataset_path = '../input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv'
stroke_df = pd.read_csv(stroke_dataset_path)

#Clean Empty Data and Convert gender
filtered_stroke_df = stroke_df.replace(['Male', 'Female', 'Other'],[0,1,np.nan]).dropna(axis=0)
filtered_stroke_df = filtered_stroke_df.replace(['Male', 'Female', 'Other'],[0,1,np.nan]).dropna(axis=0)

#Define the target
y = filtered_stroke_df.stroke

#Define the features
features = ['age','gender', 'hypertension', 'heart_disease','bmi', 'avg_glucose_level']

X = filtered_stroke_df[features]
X = scale(X)

#Split data on train and validation data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=10)

clf = MLPClassifier(solver='sgd', alpha=1e-5, hidden_layer_sizes=(6, 4, 1), random_state=1)
clf.fit(train_X, train_y)
clf.score(val_X, val_y)
```
