Having the knowledge in advance, which factor causes customers to leave their credit card services, and the ability to predict which customer going to get churned gives the bank the ability for offering better conditions for the customer and keep him as a satisfied customer. This dataset involves different features about each customer including salary, marital status, credit card limit, credit card category, etc. Initially, data exploration was shown to study the customers and then a model was train to predict churning customers.
```python
bank_data['Attrition_Flag'].value_counts()
bank_data['Attrition_Flag'].replace({'Existing Customer':0,'Attrited Customer':1},inplace=True)
bank_data.head()
```

```python
existing_customer=bank_data['Attrition_Flag']
existing_customer=bank_data[existing_customer==0]

attrited_customer=bank_data['Attrition_Flag']
attrited_customer=bank_data[attrited_customer==1]

num_of_existing=existing_customer['Attrition_Flag'].count()
num_of_attrited=attrited_customer['Attrition_Flag'].count()

print(f' The number of attrited customers is:{num_of_attrited}, while the number of existing customers is:{num_of_existing} ')
```

It can be shown that there are much more data about existing customers than attrited customers. Preventing misclassifications of the model requires us to do upsampling.

```python
df_minority_upsampled = resample(attrited_customer, 
                                 replace=True,    
                                 n_samples=existing_customer['Attrition_Flag'].count()) 
df_upsampled = pd.concat([existing_customer, df_minority_upsampled])
df_upsampled.info()
```

```python
df_upsampled['Gender'].replace({'M':0,'F':1},inplace=True)
df_upsampled.replace(to_replace='Unknown',value=np.nan,inplace=True)
df_upsampled=pd.get_dummies(df_upsampled,columns=['Education_Level','Marital_Status','Income_Category','Card_Category'])
```

```python
df_upsampled=df_upsampled[['Customer_Age', 'Gender', 'Dependent_count',
       'Months_on_book', 'Total_Relationship_Count', 'Months_Inactive_12_mon',
       'Contacts_Count_12_mon', 'Credit_Limit', 'Total_Revolving_Bal',
       'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1', 'Total_Trans_Amt',
       'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1', 'Avg_Utilization_Ratio',
       'Education_Level_College', 'Education_Level_Doctorate',
       'Education_Level_Graduate', 'Education_Level_High School',
       'Education_Level_Post-Graduate', 'Education_Level_Uneducated',
       'Marital_Status_Divorced', 'Marital_Status_Married',
       'Marital_Status_Single', 'Income_Category_$120K +',
       'Income_Category_$40K - $60K', 'Income_Category_$60K - $80K',
       'Income_Category_$80K - $120K', 'Income_Category_Less than $40K',
       'Card_Category_Blue', 'Card_Category_Gold', 'Card_Category_Platinum',
       'Card_Category_Silver','Attrition_Flag']]
df_upsampled.head()
```

 <div style="color:black;
           display:fill;
           border-radius:5px;
           font-size:300%;
           font-family: Times New Roman, Times, serif;
    letter-spacing:0.5px"><b>Training the model</b>

<p style="padding: 10px;color:black;font-size:40%;font-family: Times New Roman, Times, serif;">
In this section I will replace the unknown data with the median of each column, to avoid outliers, then I'll split the data to train, test and validation. Next, training will be taken using the Random Forest model.
</p>

```python
X=df_upsampled.iloc[:,:-1].values
Y=df_upsampled.iloc[:,-1].values
imputer = SimpleImputer(missing_values=np.nan, strategy='median')
X= imputer.fit_transform(X)
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.25)
x_train,x_valid,y_train,y_valid=train_test_split(x_train,y_train,test_size=0.1)

forest=RandomForestClassifier()
train=forest.fit(x_train,y_train)
acc=train.score(x_valid,y_valid)
print(f' The model accuracy on the validation data is :{acc:.2%}')
```

 <div style="color:black;
           display:fill;
           border-radius:5px;
           font-size:300%;
           font-family: Times New Roman, Times, serif;
    letter-spacing:0.5px"><b>Predications</b>

<p style="padding: 10px;color:black;font-size:40%;font-family: Times New Roman, Times, serif;">
Here the model will try to predict the results of the test data(i.e., the data it didn't see) and measure its accuracy.
</p>

```python
pred=train.predict(x_test)
pred=pred.reshape(-1,1)
y_test=y_test.reshape(-1,1)
total_predications=pred-y_test
val,freq=np.unique(total_predications,return_counts=True)
accuracy=freq[1]/np.sum(freq)
print(f' The model accuracy on the validation data is :{accuracy:.2%}')
```
