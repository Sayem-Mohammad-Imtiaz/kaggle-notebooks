```python
from plotly import*
from plotly.offline import iplot
```

First importing data;

```python
data=pd.read_csv("../input/world-happiness-report-2021/world-happiness-report-2021.csv")
```

```python
data.info()
```

```python
data.head(11)
```

Make some columns available shape;


```python
data=data.rename(columns={'Country name': 'country','Ladder score': 'score','Logged GDP per capita': 'gdp'})
```

```python
data.head()
```

```python
x = data.country
plt.subplots(figsize=(8,8))
wordcloud = WordCloud(
                          background_color='white',
                          width=512,
                          height=384
                         ).generate(" ".join(x))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('graph.png')

plt.show()
```

Ploting a bubble chart depending on the happiness score of the countries

```python
# data preparation
d = [
    {
        'y': data.gdp,
        'x': data.index,
        'mode': 'markers',
        'marker': {
            'color': 'green',
            'size': data.score,
            'showscale': True
        },
        "text" :  data.country    
    }
]
iplot(d)
```
