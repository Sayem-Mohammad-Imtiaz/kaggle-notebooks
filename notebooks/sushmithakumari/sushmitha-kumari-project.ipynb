```python
sc_cols=[2,3,9,11,15,16]
```

```python
trans_data=make_column_transformer((StandardScaler(),sc_cols),
                                  remainder='passthrough')
```

```python
set_config(display="diagram")

pipe=make_pipeline(trans_data,model)

pipe
```
