# Models construction

```python
wcss=[]

for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(df[['Age','Spending Score (1-100)']])
    wcss.append(kmeans.inertia_)

#Visualizing the ELBOW method to get the optimal value of K 
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# the last elbow is at k=4
kmeansmodel = KMeans(n_clusters= 4, init='k-means++', random_state=42)
y_kmeans= kmeansmodel.fit_predict(df[['Age','Spending Score (1-100)']])


plt.figure(figsize=(14,7))
plt.title('Age -  Spending Score Segmentation') 
plt.xlabel('Age') 
plt.ylabel('Spending Score')
colors = ['C0', 'C1','C2','C3','C4']
for i in range(len(np.unique(y_kmeans))):
    plt.scatter(df.loc[y_kmeans == i, 'Age'], 
                df.loc[y_kmeans == i, 'Spending Score (1-100)'], 
                s = 100, c=colors[i], label = 'Cluster {}'.format(i+1))
plt.scatter(kmeansmodel.cluster_centers_[:, 0], 
            kmeansmodel.cluster_centers_[:, 1], 
           s = 200, marker='x', c = 'black', linewidth=6, label = 'Centroids')
plt.legend()
plt.show()
```

## Model 1 Interpretation
Cluster 3, green, shows us that people between 18 and 40 years old are the best target

```python
wcss=[]

for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(df[["Annual Income (k$)","Spending Score (1-100)"]])
    wcss.append(kmeans.inertia_)
    
#Visualizing the ELBOW method to get the optimal value of K 
plt.plot(range(1,11), wcss)
plt.title('The Elbow Method')
plt.xlabel('no of clusters')
plt.ylabel('wcss')
plt.show()
```

```python
# the last elbow is at k=5
kmeansmodel = KMeans(n_clusters= 5, init='k-means++', random_state=42)
y_kmeans= kmeansmodel.fit_predict(df[['Annual Income (k$)','Spending Score (1-100)']])


plt.figure(figsize=(14,7))
plt.title('Annual Income -  Spending Score Segmentation')
plt.xlabel('Annual Income') 
plt.ylabel('Spending Score')
colors = ['C0', 'C1','C2','C3','C4']
for i in range(len(np.unique(y_kmeans))):
    plt.scatter(df.loc[y_kmeans == i, 'Annual Income (k$)'], 
                df.loc[y_kmeans == i, 'Spending Score (1-100)'], 
                s = 100, c=colors[i], label = 'Cluster {}'.format(i+1))
plt.scatter(kmeansmodel.cluster_centers_[:, 0], 
            kmeansmodel.cluster_centers_[:, 1], 
            s = 200, marker='x', c = 'black', linewidth=6, label = 'Centroids')

plt.legend()
plt.show()
```

## Model 1 Interpretation
Cluster 1, blue,   average earning and average spending (target 3) <br>
Cluster 2, orange, high earning but low spending<br>
Cluster 3, green,  low earning and low spending<br>
Cluster 4, red,    low earning and high spending (target 2)<br>
Cluster 5, purple, high earning and high spending (target 1)<br>
