# 2) Matplotlib

```python
# Line plot

# Black rating and white rating
data.black_rating.plot(kind='line', color='b', label='Black Rating', linewidth=1, alpha=0.5, grid = True, linestyle=':')
data.white_rating.plot(color='r', label='White Rating', linewidth=1, alpha=0.5, grid=True, linestyle='-.')
plt.xlabel('mac id')
plt.ylabel('degerleri')
plt.show()

```

```python
# Scatter Plot

plt.scatter(data.black_rating, data.white_rating, color='g', alpha=0.5)
plt.xlabel('Black Rating')
plt.ylabel('White Rating')
plt.title('Black Rating and White Rating Scatter Plot')
plt.show()

# Burda black'in rating'i artarsa white'in da artar.
```

```python
# histogram plot

data.white_rating.plot(kind = 'hist', bins=50,figsize= (18,18))
plt.show()
```
