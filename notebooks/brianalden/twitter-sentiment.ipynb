**Preparing the data**

```python
dforig = pd.read_csv('/kaggle/input/twitter-airline-sentiment/Tweets.csv')
df = dforig.copy()
df = df.drop(['retweet_count', 'tweet_coord', 'tweet_created', 'tweet_location','name','user_timezone','negativereason_gold','tweet_id','negativereason','airline_sentiment_confidence','airline','name','negativereason','negativereason_confidence','airline_sentiment_gold','tweet_location','tweet_coord'], axis = 1)
df
```

**Data  preprocessing**

```python
from sklearn.model_selection import train_test_split
import nltk
from nltk.corpus import stopwords
from nltk.classify import SklearnClassifier
from wordcloud import WordCloud,STOPWORDS
import string
from nltk.stem import PorterStemmer
import matplotlib.pyplot as plt
```

```python
df=df[['text','airline_sentiment']]
df.head()
```

```python
df['airline_sentiment'].value_counts()
```

```python
df = df[df['airline_sentiment'] != 'neutral'] # see only pos/neg
df['airline_sentiment'].value_counts() # updated df
```

```python
df.info()
```

**Count Vectorizer**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words = 'english')

dtm = cv.fit_transform(df['airline_sentiment'])
```

**Justify X&Y Variable**

```python
X = dtm
Y = df['text']
```

**Testing(Training and predicting Data)**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 1)

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()

# Train
rfc.fit(X_train, Y_train)

# Predict
Y_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, classification_report
print(confusion_matrix(Y_test, Y_pred))
print(" ")
print(accuracy_score(Y_test, Y_pred))
print(" ")
print(classification_report(Y_test, Y_pred))
```
