```python
import numpy as ny
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('../input/insurance/insurance.csv')
df.head()
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
plt.figure(figsize=(5,6))
sns.countplot(x="region",hue='sex',data=df)
```

```python
dum = pd.get_dummies (df.sex)
merged = pd.concat([df,dum],axis=1)
df = merged.drop(['sex'],axis=1)
```

```python
dum = pd.get_dummies (df.smoker)
merged = pd.concat([df,dum],axis=1)
df = merged.drop(['smoker'],axis=1)
```

```python
dum = pd.get_dummies (df.region)
merged = pd.concat([df,dum],axis=1)
df = merged.drop(['region'],axis=1)
```

```python
df.head()
```

```python
x = df.drop(['charges','northeast','northwest','southeast','southwest'],1)
x.head()
```

```python
y = df['charges']
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test =train_test_split(x,y,test_size=0.33)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
pred = model.predict(X_test)
```

```python
model.score(X_test,y_test)
```
