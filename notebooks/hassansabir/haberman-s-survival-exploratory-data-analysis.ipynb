```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
haberman=pd.read_csv("/kaggle/input/habermans-survival-data-set/haberman.csv")
```

```python
haberman=haberman.rename(columns={"30":"Age","64":"Op_Year","1":"axil_nodes","1.1":"Surv_status"})
haberman["Surv_status"] = haberman["Surv_status"].astype(str)
haberman["Surv_status"].replace({"1": "yes", "2": "no"}, inplace=True)
haberman
```

number of features=4(include class) <br>
number of class =1 (Surv_status) <br>
Surv_status(class) <br>
yes = the patient survived 5 years or longer <br>
no = the patient died within 5 year 

```python
haberman["Surv_status"].value_counts()
```

# OBJECTIVE
Survival of patients who had undergone surgery for breast cancer

```python
sns.pairplot(haberman,hue="Surv_status")
```

# HISTOGRAM , PDF ,CDF

```python
sns.set_style('whitegrid')
sns.FacetGrid(data=haberman,hue="Surv_status",height=5).map(sns.distplot,"Age").add_legend()
plt.show()
```

It shows that the patient which has Age  between 43 and 53 has less chance of survival while the patient has Age between 20 and 43 has more chance of survival

```python
sns.FacetGrid(data=haberman,hue="Surv_status",size=6)\
    .map(sns.distplot,"axil_nodes")\
    .add_legend()
plt.show()
```

if patient have axil_nodes <= 2 it has more chance of survival <br>
if patient have axile_nodes > 2 than it has very less chance of suvival

```python
survival=haberman.loc[haberman["Surv_status"]=="yes"]
dead=haberman.loc[haberman["Surv_status"]=="no"]
counts,bin_edges=np.histogram(survival["axil_nodes"],density=True)
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
y=[pdf,cdf]
labels=["survived pdf","survived cdf"]
for y_arr, label in zip(y, labels):
    plt.plot(bin_edges[1:], y_arr, label=label)


counts,bin_edges=np.histogram(dead["axil_nodes"],density=True)
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
y=[pdf,cdf]
labels=["dead pdf","dead cdf"]
for y_arr, label in zip(y, labels):
    plt.plot(bin_edges[1:], y_arr, label=label)
plt.xlabel("Axile_nodes")
plt.legend()
plt.show()
```

90% survived people have axile_nodes < 6 and 63% dead people have axile_nodes >= 6

```python
counts,bin_edges=np.histogram(survival["Age"],density=True)
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
y=[pdf,cdf]
labels=["survived pdf","survived cdf"]
for y_arr, label in zip(y, labels):
    plt.plot(bin_edges[1:], y_arr, label=label)


counts,bin_edges=np.histogram(dead["Age"],density=True)
pdf=counts/(sum(counts))
cdf=np.cumsum(pdf)
y=[pdf,cdf]
labels=["dead pdf","dead cdf"]
for y_arr, label in zip(y, labels):
    plt.plot(bin_edges[1:], y_arr, label=label)

plt.xlabel("Age")
plt.legend()
plt.show()
```

20% to 65% of dead patients have age between 43 to 55 <br>
30% survived people age is < 43 and 60% to 70% survived people age is between 55 to 61
