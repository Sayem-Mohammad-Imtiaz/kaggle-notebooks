```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
data= pd.read_csv("../input/haberman.csv",names=['Age', 'Year', 'Axillary nodes dect', 'Surv_status'])
```

```python
a=data.columns
```

```python
for i in range(len(a)-1):
    for j in range(i+1,len(a)-1):
        sn.FacetGrid(data, hue="Surv_status",height=5)\
        .map(plt.scatter,a[i],a[j]).add_legend();
        plt.show()
```

## Observation

1. All the data points are mixed and cannot be seperated using a linear line or a curve

```python
sn.set_style("whitegrid")
sn.pairplot(data,hue="Surv_status",height=3)
plt.show()
```

## Observation:

 1. Even the pair plots do not show any linearly seperable data, so we cannot come to a conclusion

```python
data_0=data.loc[data["Surv_status"]==1];
data_1=data.loc[data["Surv_status"]==2];
for i in range(0,len(a)-1):
    plt.plot(data_0[a[i]],np.zeros_like(data_0[a[i]]),'bo')
    plt.plot(data_1[a[i]],np.zeros_like(data_1[a[i]]),'r^')
    plt.show()
```

## Observation

1. The distribution is mixed and a proper relation cannot be bought

```python
for i in range(0,len(a)-1):
    sn.FacetGrid(data, hue="Surv_status",height=5)\
    .map(sn.distplot, a[i]).add_legend()
    plt.show()
```

## Observation:

According to the overlapped area, area of axil_nodes<age~Op_year
So a little bit of differentiation can be done using axil_nodes

```python
for i in range(0,len(a)-1):
    counts, bin_edges = np.histogram(data_0[a[i]],bins=10, density=True)
    pdf = counts/(sum(counts))
    print(pdf)
    cdf = np.cumsum(pdf)
    plt.plot(bin_edges[1:],pdf)
    plt.plot(bin_edges[1:],cdf)
    #plt.show()
    counts, bin_edges = np.histogram(data_1[a[i]],bins=10, density=True)
    pdf = counts/(sum(counts))
    print(pdf)
    cdf = np.cumsum(pdf)
    plt.plot(bin_edges[1:],pdf)
    plt.plot(bin_edges[1:],cdf)
    plt.show()
```

## Observation:

The PDF and CDF curves of Age and Op_year both are mostly on the same line
The PDF and CDF curve of Axil_nodes show that there is a 0.8 probability that the person survives when axil_node is less

## Mean:

We are calculating the mean

```python
for i in range(0,len(a)-1):
    print(np.mean(data_0[a[i]]))
    print(np.mean(data_1[a[i]]))
    print(np.mean(data[a[i]]))
    print()
```

## Observation

The means of age,op_age data 1 and data 2 are nearly equal to data
The mean of axil_node is a little bit different

## Median

Here we will be calculating the median

```python
for i in range(0,len(a)-1):
    print(np.median(data_0[a[i]]))
    print(np.median(data_1[a[i]]))
    print(np.median(data[a[i]]))
    print()
```

## Observation:

The median of age and op_age are same
The axil_node are different. So it can be used to perform the classification

## Standard deviation

```python
for i in range(0,len(a)-1):
    print(np.std(data_0[a[i]]))
    print(np.std(data_1[a[i]]))
    print(np.std(data[a[i]]))
    print()
```

## Observation

The standard deviation is same for age, op_age but axil_node is different

## Percentile

Here we calculate the percentile

```python
for i in range(0,len(a)-1):
    print(np.percentile(data_0[a[i]],np.arange(0, 100, 25)))
    print(np.percentile(data_1[a[i]],np.arange(0, 100, 25)))
    print(np.percentile(data[a[i]],np.arange(0, 100, 25)))
    print()
```

## Observation:
