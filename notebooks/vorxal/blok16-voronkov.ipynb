3) Создать новый столбец, в котором разместить разницу между максимумом и минимумом

```python
#Скопируем исходные данные, чтобы не портить их
new_data = data

new_data["diff"] = new_data.max(axis = 1) - new_data.min(axis = 1)
print(new_data)
```

4) Разбить данные на несколько блоков (не менее 3, приблизительно равных по размеру) по некоторому критерию (критерий определить самостоятельно)

```python
#Т.к. у меня по зданию присутствует дата, то воспользуемся разбиением по дате и раздели данные на 4 части: зима, весна, лето, осень
#Я около 2.5 часов пытался разделить данные и смог только с помощью регулярных выражений. Классическим функциям питона в моём исполнении не удалось побороть данную задачу
winter_df = new_data[new_data['Country/Region'].str.match('^(1|2|12)/')]
spring_df = new_data[new_data['Country/Region'].str.match('^(3|4|5)/')]
summer_df = new_data[new_data['Country/Region'].str.match('^(6|7|8)/')]
fall_df = new_data[new_data['Country/Region'].str.match('^(9|10|11)/')]
print(winter_df)
print(spring_df)
print(summer_df)
print(fall_df)
```

5) Привести круговую диаграмму для проделанного разбиения.

```python
#Знаю, что можно визуализировать данную задачу с помощью Pandas, но лучше сделаю это с помощью matplotlib
#Визуализировать будем данные по России
import matplotlib.pyplot as plt
values = [sum(winter_df["Russia"]),
          sum(spring_df["Russia"]),
          sum(summer_df["Russia"]),
          sum(fall_df["Russia"])]
names = ["Winter", "Spring", "Summer", "Fall"]
fig1, ax1 = plt.subplots()
ax1.pie(values, labels=names)
plt.show()
```

6) В зависимости от особенностей набора данных для каждого блока построить график (гистограмму, ...)

```python
winter_df_x = winter_df["Country/Region"].values
winter_df_y = winter_df["diff"].values

fig, ax = plt.subplots()

ax.bar(winter_df_x, winter_df_y)
ax.set_facecolor('seashell')
ax.set_xticklabels(winter_df_x, rotation = 90)
fig.set_facecolor('floralwhite')
fig.set_figwidth(12)    #  ширина Figure
fig.set_figheight(6)    #  высота Figure
plt.show()

spring_df_x = spring_df["Country/Region"].values
spring_df_y = spring_df["diff"].values

fig, ax = plt.subplots(figsize=(21,3), dpi=1000)

ax.bar(spring_df_x, spring_df_y)
ax.set_facecolor('seashell')
ax.set_xticklabels(spring_df_x, rotation = 90)
fig.set_facecolor('floralwhite')
fig.set_figwidth(12)
fig.set_figheight(6) 
plt.show()

summer_df_x = summer_df["Country/Region"].values
summer_df_y = summer_df["diff"].values

plt.rcParams["figure.figsize"] = (20,3)
plt.plot(summer_df_x, summer_df_y) 
plt.xlabel('Ось X')
plt.ylabel('Ось Y')
plt.title('Заголовок графика')
plt.show();
```

7) Сгруппировать данные (GroupBy) по некоторому признаку и сохранить результаты в новые таблицы. Для каждой новой таблицы провести сортировку по сложному ключу, состоящему из нескольких признаков. Для каждой таблицы своя сортировка.

```python
#С рпедставленными данными довольно сложно выбрать признак группировки, поэтому отобразим данное задание с помощью группировки по разнице 
# и сортировки по Вьетнаму(для показательности) и Украины
gk = new_data.groupby('diff')
gk.first()
a = gk.get_group(2.0)
a.sort_values(by=["Vietnam","Ukraine"],key=lambda x:x)
```

```python
#А тут у нас сортировочка по Китаю и Вьетнаму, логично предположить, что в Китае в тот момент было больше заболеваний, поэтому выведем отсортированные данные по Китаю
b = gk.get_group(3.0)
b.sort_values(by=["China","Vietnam"],key=lambda x:x)["China"]
```
