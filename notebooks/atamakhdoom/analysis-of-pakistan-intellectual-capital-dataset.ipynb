```python
data = pd.read_csv("../input/pakistanintellectualcapitalcs/Pakistan Intellectual Capital - Computer Science - Ver 1.csv", encoding = "ISO-8859-1")
data.head()
```

```python
data.info()
```

```python
my_data = data
my_data = my_data.drop('S#', axis=1)
my_data.head()
```

```python
# Renaming columns names for convinience
my_data = my_data.rename(columns={
    "Teacher Name": "teacher_name",
    "University Currently Teaching": "current_university",
    "Department": "department",
    "Province University Located": "province",
    "Designation": "designation",
    "Terminal Degree": "degree",
    "Graduated from": "graduated_from",
    "Country": "country",
    "Year": "year",
    "Area of Specialization/Research Interests": "specialization",
    "Other Information": "other_information"
})
my_data.head()  # Displaying initial five rows
```

```python
my_data['department'].value_counts()
```

```python
# REPLACE DEPARTMENT NAME IN DATASET FOR DATA UNIQUENESS
my_data['department'] = my_data['department'].replace('Computer Science', 'Computer Sciences')
my_data['department'] = my_data['department'].replace('Department of Computer Systems & Software Engineering', 'Computer Science and Software Engineering')
my_data['department'] = my_data['department'].replace('Computer Science & Engineering', 'Computer Science and Engineering')
my_data['department'] = my_data['department'].replace('CS/IT', 'Computer Science & Information Technology')
my_data['department'] = my_data['department'].replace('CS & IT', 'Computer Science & Information Technology')
my_data['department'] = my_data['department'].replace('Computer Science and IT', 'Computer Science & Information Technology')
my_data['department'] = my_data['department'].replace('Computer Science & IT', 'Computer Science & Information Technology')
my_data['department'] = my_data['department'].replace('Computer Science and Information Technology', 'Computer Science & Information Technology')
my_data['department'] = my_data['department'].replace('Computing & Information Sciences', 'Computer and Information Sciences')
```

```python
my_data['department'].value_counts()
```

```python
my_data['degree'].unique()
```

```python
my_data['degree'] = my_data['degree'].replace("Ph.D","PhD")
my_data['degree'] = my_data['degree'].replace("Phd","PhD")
my_data['degree'] = my_data['degree'].replace("Ph.D (Scholar)","PhD")
my_data['degree'] = my_data['degree'].replace("Ph.D(Scholar)","PhD")

my_data['degree'] = my_data['degree'].replace("Mphil","M.Phil")
my_data['degree'] = my_data['degree'].replace("MPhil","M.Phil")
my_data['degree'] = my_data['degree'].replace("M.Phil leading to PhD","M.Phil")

my_data['degree'] = my_data['degree'].replace("MSc","M.Sc")
my_data['degree'] = my_data['degree'].replace("MSC","M.Sc")

my_data['degree'] = my_data['degree'].replace(" MCS","MCS")
my_data['degree'] = my_data['degree'].replace("Mcs","MCS")

my_data['degree'] = my_data['degree'].replace("MS (Scholar)","MS")
my_data['degree'] = my_data['degree'].replace("M.S","MS")

my_data['degree'] = my_data['degree'].replace("M.E","ME")

my_data['degree'] = my_data['degree'].replace("BSC","B.Sc")
my_data['degree'] = my_data['degree'].replace("BSc","B.Sc")

my_data['degree'] = my_data['degree'].replace("B.E","BE")

my_data['degree'] = my_data['degree'].replace("PostDoc","Post Doc")
my_data['degree'] = my_data['degree'].replace("Master","Masters")
```

```python
my_data['degree'].unique()
```

```python
my_data["country"].unique()
```

```python
my_data["country"] = my_data["country"].replace(" USA", "USA")
my_data["country"] = my_data["country"].replace(" New Zealand", "New Zealand")
my_data["country"] = my_data["country"].replace(" Germany", "Germany")
my_data["country"] = my_data["country"].replace(" Sweden", "Sweden")
```

```python
my_data["country"].unique()
```

```python
my_data.department.groupby(my_data.department).count().sort_values(ascending=False).plot(kind="bar")
```

```python
my_data["degree"].groupby(my_data["degree"]).count().sort_values(ascending=False)[:20].plot(kind="bar")
```
