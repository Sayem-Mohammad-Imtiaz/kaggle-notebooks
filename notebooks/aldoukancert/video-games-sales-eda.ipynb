## **-Categorical Variable**

```python
def bar_plot(variable):
    var = vg_df[variable]
    
    varValue = var.value_counts()
    
    varValue = var.value_counts()
    
    plt.figure(figsize=(30,10))
    plt.bar(varValue.index,varValue)
    plt.xticks(varValue.index,varValue.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
    print("{}: \n {}".format(variable,varValue))
```

# **4.Outlier Detection**

```python
def detect_outliers(vg_df,features):
    outlier_indices = []
    
    for c in features:
        # 1st quartile
        Q1 = np.percentile(vg_df[c],25)
        # 3rd quartile
        Q3 = np.percentile(vg_df[c],75)
        # IQR
        IQR = Q3 - Q1
        # Outlier step
        outlier_step = IQR * 1.5
        # detect outlier and their indeces
        outlier_list_col = vg_df[(vg_df[c] < Q1 - outlier_step) | (vg_df[c] > Q3 + outlier_step)].index
        # store indeces
        outlier_indices.extend(outlier_list_col)
    
    outlier_indices = Counter(outlier_indices)
    multiple_outliers = list(i for i, v in outlier_indices.items() if v > 2)
    
    return multiple_outliers
```

```python
vg_df.loc[detect_outliers(vg_df,["NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"])]
```

# **5.Visualization of Some Data**

<a id="9"></a><br>
## **-Bar Plot**

```python
vg_df.head()
```

```python
vg_df.info()
```

```python
genre_list=list(vg_df["Genre"].unique())
NA_Sales1 = []      
EU_Sales1 = []    
JP_Sales1 = []     
Other_Sales1 = []  
Global_Sales1 = []
for i in genre_list:
    x = vg_df[vg_df["Genre"]==i]
    NA_Sales1.append(sum(x.NA_Sales)/len(x))     
    EU_Sales1.append(sum(x.EU_Sales)/len(x))  
    JP_Sales1.append(sum(x.JP_Sales)/len(x)) 
    Other_Sales1.append(sum(x.Other_Sales)/len(x)) 
    Global_Sales1.append(sum(x.Global_Sales)/len(x))
    
#visualization
f,ax=plt.subplots(figsize=(9,12))
sns.barplot(x=NA_Sales1,y=genre_list,color="blue",alpha=0.5,label="NA Sales")
sns.barplot(x=EU_Sales1,y=genre_list,color="purple",alpha=0.6,label="EU Sales")
sns.barplot(x=JP_Sales1,y=genre_list,color="red",alpha=0.7,label="JP Sales")
sns.barplot(x=Other_Sales1,y=genre_list,color="yellow",alpha=0.8,label="Other Sales")
sns.barplot(x=Global_Sales1,y=genre_list,color="green",alpha=0.4,label="Global Sales")

ax.legend(loc="lower right",frameon=True)
ax.set(xlabel="Rates of Sales",ylabel="Games of Genre",title="Sales Rates of Game Genres in The Region")
plt.show()
```
