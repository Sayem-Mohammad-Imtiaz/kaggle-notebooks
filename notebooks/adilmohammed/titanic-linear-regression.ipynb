```python
data = pd.read_csv("../input/titanic/train_and_test2.csv")
data.head(10)
```

```python
data.shape
```

```python
null = data.isnull().sum()
null
```

```python
data = data.dropna(axis = 0)
```

```python
train_data = data[0:900]
train_data.shape
```

```python
test_data = data[900:]
test_data.shape
```

```python
features = ["Age", "Sex", "Embarked"]
train_input = train_data[features]
```

```python
train_output = train_data['2urvived']
```

```python
features = ["Age", "Sex", "Embarked"]
test_input = test_data[features]
```

```python
test_output = test_data['2urvived']
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(train_input, train_output)
predictions = linear_reg.predict(test_input)
```

```python
from sklearn.metrics import mean_squared_error
error = mean_squared_error(test_output, predictions)
error
```
