```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import *
```

**Importing the csv whith the datas**

```python
df = pd.read_csv('/kaggle/input/credit-card-customers/BankChurners.csv')
df
```

**Creating the dependent variable using "Attrition_Flag" with 1 when "Existing Customer" and 0 when "Attrited Customer"**

```python
Y = pd.get_dummies(df["Attrition_Flag"])
Y = Y["Existing Customer"]
```

**Showing the relations of the variables what will be used**

```python
fig, axs = plt.subplots(2, 2)
plt.figure(figsize=(10,10))
axs[0,0].scatter(df["Months_on_book"],df["Total_Trans_Amt"], c= Y)
plt.figure(figsize=(10,10))
axs[1,0].scatter(df["Total_Trans_Amt"],df["Avg_Open_To_Buy"], c= Y)
plt.figure(figsize=(10,10))
axs[1,1].scatter(df["Avg_Open_To_Buy"],df["Total_Trans_Ct"], c= Y)
plt.figure(figsize=(10,10))
axs[0,1].scatter(df["Total_Trans_Ct"],df["Months_Inactive_12_mon"], c= Y)
```

**Spliting the samples**

```python
X = df[["Months_on_book", "Total_Trans_Amt", "Avg_Open_To_Buy", "Total_Trans_Ct", "Months_Inactive_12_mon"]]
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y,test_size=0.33,random_state=42)
```

**Creating the model of classification**

```python
clf2 = tree.DecisionTreeClassifier(criterion = "entropy", splitter = "best", max_depth = 10)
clf2 = clf2.fit(X_train, Y_train)
```

**Evaluating the model outsample**

```python
print("Accuracy : ", metrics.accuracy_score(Y_test, clf2.predict(X_test)) *  100)
print("Recall : ", metrics.recall_score(Y_test, clf2.predict(X_test)) *  100)
print("Precision : ", metrics.precision_score(Y_test, clf2.predict(X_test)) *  100)
print(metrics.confusion_matrix(Y_test, clf2.predict(X_test)))
print(metrics.classification_report(Y_test, clf2.predict(X_test)))
```

**Creating the model whith all sample**

```python
clf = tree.DecisionTreeClassifier(criterion = "entropy", splitter = "best", max_depth = 10)
clf = clf2.fit(X, Y)
```

**Avaluating th model whith all sample**

```python
print("Accuracy : ", metrics.accuracy_score(Y, clf.predict(X)) *  100)
print("Recall : ", metrics.recall_score(Y, clf.predict(X)) *  100)
print("Precision : ", metrics.precision_score(Y, clf.predict(X)) *  100)
print(metrics.confusion_matrix(Y, clf.predict(X)))
print(metrics.classification_report(Y, clf.predict(X)))
```
