# Medical Cost Personal Dataset
**Columns**

age: age of primary beneficiary

sex: insurance contractor gender, female, male

bmi: Body mass index, providing an understanding of body, weights that are relatively high or low relative to height,objective index of body weight (kg / m ^ 2) using the ratio of height to weight, ideally 18.5 to 24.9

children: Number of children covered by health insurance / Number of dependents

smoker: Smoking

region: the beneficiary's residential area in the US, northeast, southeast, southwest, northwest.

charges: Individual medical costs billed by health insurance

In notebook I am trying determine predictive variables by analysing the dataset visualy

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('../input/insurance/insurance.csv')
df.head()
```

# Finding correlation with Charges

# BMI

```python
sns.scatterplot(x='bmi',y='charges',data=df)
sns.lmplot(x='bmi',y='charges',data=df)
```

Bmi seems to be an important predictor. Lets test it with other parameter.

```python
sns.scatterplot(x='bmi',y='charges',hue='smoker',data=df)
sns.lmplot(x='bmi',y='charges',hue='smoker',data=df)
```

```python
sns.scatterplot(x='bmi',y='charges',hue='region',data=df)
sns.lmplot(x='bmi',y='charges',hue='region',data=df)
```

```python
sns.scatterplot(x='bmi',y='charges',hue='sex',data=df)
sns.lmplot(x='bmi',y='charges',hue='sex',data=df)
```

For now ,BMI seems to a good predictor along with it smoker seems to be good predicator along with it sex and region seems to be a mild predictor.
