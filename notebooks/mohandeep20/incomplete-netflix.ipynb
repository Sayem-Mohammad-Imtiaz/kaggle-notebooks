### For a clear understanding of the ratings, we will replace the current rating with age group.

```python
ratings = {
    'TV-PG': '12-14+',
    'TV-MA': '18+',
    'TV-Y7-FV': ' 7+',
    'TV-Y7': ' 7+',
    'TV-14': '12-14+',
    'R': '18+',
    'TV-Y': ' 0+ (ALL)',
    'NR': '18+',
    'PG-13': '12-14+',
    'TV-G': ' 0+ (ALL)',
    'PG': '12-14+',
    'G': ' 0+ (ALL)',
    'UR': '18+',
    'NC-17': '18+'
}
data_india.rating = data_india.rating.replace(ratings)
```
