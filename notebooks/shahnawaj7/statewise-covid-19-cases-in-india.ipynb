```python
cor=df.corr().transpose().style.set_properties(**{'background-color':'black',
                                  'color':'gold',
                                  })
cor
```

```python
cor=df.corr()
sns.heatmap(cor, xticklabels = cor.columns, yticklabels = cor.columns, annot = True)
```

```python
sns.catplot(x='State/UTs', y="Deaths", kind="bar",
            data=df.head(7),aspect=2)
```

```python
sns.catplot(x='State/UTs', y="Total Cases", kind="boxen",
            data=df.head(7),aspect=2)
```

```python
sns.catplot(x='State/UTs', y="Discharged", kind="point",
            data=df.head(7),aspect=2)
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,x_vars=['Total Cases','Active'],y_vars='Deaths',hue='State/UTs')
```

```python
sns.relplot(x='Total Cases',y='Deaths',data=df,hue='State/UTs')
```

```python
sns.displot(data=df['Deaths'])
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['State/UTs'], y=df.Active, 
                        mode='lines+markers'))
```

```python
fig = px.scatter(df, x="State/UTs", y="Total Cases", color="Active",)

fig.show()
```

```python
fig = px.bar_polar(
df, r="Death Ratio (%)", theta="State/UTs", color="Deaths",
                   color_discrete_sequence= px.colors.sequential.Plasma_r,
                   title="Deaths vs Death Ratio Visualisation"
)
fig.show()
```

```python
fig = px.scatter_3d(
df, x='State/UTs', z='Active', y='Discharge Ratio (%)',color = 'Total Cases',
                          opacity=0.5, width=800, height=900
)
fig.show()
```

```python
fig = px.scatter_polar(
df, r="Active Ratio (%)", theta="State/UTs", color="Deaths",
                   color_discrete_sequence= px.colors.sequential.Plasma_r,
                   title="Deaths vs Active Ratio Visualisation"
)
fig.show()
```
