```python
#libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#data
data = pd.read_csv("../input/mushroom-classification/mushrooms.csv")
data_origin=data.copy()
data.head()
```

```python
data.describe()
```

```python
#pre-processing
data.nunique()
```

```python
data=data.drop(['veil-type','class'],axis=1)
```

```python

#oneHotencoder
one_hot_encoded_data_origin = pd.get_dummies(data_origin)
one_hot_encoded_data = pd.get_dummies(data)
one_hot_encoded_data
```

```python
#categorical-clustering
from kmodes.kmodes import KModes

k = KModes(n_clusters=2, init='Cao', verbose=1)
clusters = k.fit_predict(one_hot_encoded_data)
predicted_labels = pd.DataFrame(clusters, columns=['predicted-label'])
```

```python
#result
predicted_labels.value_counts()
```

```python
#plot
predicted_labels.value_counts().plot.bar()
```

```python
predicted_labels.value_counts().plot.pie(autopct='%1.0f%%')
```

```python
# Plot
plt.figure(figsize=(48,40), dpi= 80)
sns.heatmap(one_hot_encoded_data.corr(), xticklabels=one_hot_encoded_data.corr().columns, yticklabels=one_hot_encoded_data.corr().columns, cmap='RdYlGn', center=0, annot=True)

# Decorations
plt.title('Correlogram of mtcars', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
labels = pd.DataFrame()
labels['target'] = one_hot_encoded_data_origin.values.reshape(119, -1).tolist()[0]
labels['prediction'] = clusters
labels.value_counts(["target", "prediction"])
```
