### 5. FILTERING FEATURES WITH NUMPY'S LOGICAL METHOD

*  Let's see the pokemons which has attack stats between 40-60

```python
df_attack_logical = np.logical_and(data["Attack"] >40,data["Attack"]< 60)
data2=data[df_attack_logical]#"df_attack_logical" has boolean values (true or false) for each sample (pokemon) in data set.
                             # It's boolean values changes according to our (between 40-60) filter.
print(data2.head()["Name"])  # We need only names so I printed first five names coloumn (feature)
```

 ### Exploring Null Data and Dropping them

* I have a huge dataset so that it is almost impossible all of features has all values.
* Before analysing data we need to make ready our dataset to analyse with some process
* It is called pre-processing
* First I need to see my data overview

```python
data.info()
```

* We have 414 non-null type_2 sample.
* If we have 800 sample dataset and there is 414 nonull values it means we have some null features (800-414=386)

```python
data["type_2"].value_counts() # It shows us the frequancies of type_2 values but there is no null values
```

```python
data["type_2"].value_counts(dropna =False) # Now we can see null values as we calculate there is 386 null values
```

* Our intention was dropping the non values in data set
* We saw that there is only one feature  has null values  which is type_2 feature
* Let's drop these bugs :)

```python
data["type_2"].dropna(inplace = True)
data["type_2"].value_counts(dropna=False)
```

* As we see there is no any null values after all 
* Let's test this with another way and finish this quest for now

```python
data["type_2"].notnull().all()
```
