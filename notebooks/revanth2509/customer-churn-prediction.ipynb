```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix ,classification_report
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
```

```python
df = pd.read_csv("../input/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df.head()
```

```python
df1 = df.drop('customerID',axis='columns')
```

```python
df1.head()
```

```python
df2 = df1[df1.TotalCharges != " "]
```

```python
df2.shape
```

```python
df2.dtypes
```

```python
df2.TotalCharges = pd.to_numeric(df2.TotalCharges)
```

```python
df2.dtypes
```

```python
df2.replace('No phone service','No',inplace=True)
df2.replace("No internet service","No",inplace=True)
```

```python
def float_to_int(df2):
    for column in df2:
        if df2[column].dtypes == object:
            df2.replace({'No':0,'Yes':1},inplace=True)
            print(f'{column}:{df2[column].unique()}')
```

```python
float_to_int(df2)
```

```python
df2.replace({'Female':0,'Male':1},inplace=True)
```

```python
df2.replace({'No':0,'Yes':1},inplace=True)
```

```python
df3 = pd.get_dummies(df2,["InternetService","Contract","PaymentMethod"])
```

```python
df3.dtypes
```

```python
plt.figure(figsize=(21,20))
plt.bar(df3['SeniorCitizen'],height=15,color=('blue'))
```

```python
x = df3.drop('Churn',axis='columns')
y = df3.Churn
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
x_train.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(26,input_shape=(26,),activation="relu"),
    keras.layers.Dense(15,activation="relu"),
    keras.layers.Dense(1,activation="sigmoid")
])
model.compile(
    loss ='binary_crossentropy',
    optimizer ='adam',
    metrics =['accuracy']
)
model.fit(x_train,y_train,epochs=500)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_prediction = model.predict(x_test)
```

```python
y_pred = []
for i in y_prediction:
    if i >= 0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
print(y_pred[:5])
print(y_test[:5])
```

```python
plt.figure(figsize=(10,9))
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('y_test')
plt.ylabel('predictions')
print(classification_report(y_test,y_pred))
```
