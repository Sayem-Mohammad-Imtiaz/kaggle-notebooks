<h1>Introduction</h1>
The aim of this kernel is to simply play around with visualization (seaborn, matplotlib and plotly) and try and find the best Pokemon for Ash to choose.  I know we all have a bias towards Pikachu but lets get Moneyball on this and remove feelings from the equation.

<h2>Importing  stuff </h2>

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import collections 

import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls

% matplotlib inline
```

```python
# lets get the dataset
poke = pd.read_csv('../input/Pokemon.csv')

#give it a little look see
poke.head()
```

```python
# the '#' feature is pointless
del poke["#"]
```

So now we want to start choosing our Pokemon, so what can be the most important stat in a Pokemon?
I'm going with Attack and Defense,  super important features

```python

fg = sns.FacetGrid(data= poke, hue='Generation', size = 8)
fg.map(plt.scatter, 'Attack','Defense').add_legend()
plt.axvline(x=75, color = 'red')
plt.axhline(y=75, color = 'red')
```

As you can see im removing all the pokemon not in the first quadrant, 75 Attack and 75 Defense is necessary man!

```python
greatattack = poke['Attack']>75
greatdefense = poke['Defense']> 75
starter = poke[greatattack & greatdefense]
```

Lets see how the dataset looks like, and while we are at it lets see how many Legendary Pokemon are there in it 

```python
fg = sns.FacetGrid(data=starter, hue='Legendary', size = 8)
fg.map(plt.scatter, 'Attack','Defense').add_legend()
```

A decent amount of Legendary Pokemon, Im removing them from the dataset, I mean if you are checking stats for a Legendary Pokemon, something is wrong with you. 

```python
notlegendary = starter['Legendary'] == False
nextstarter = starter[notlegendary]
```

Now that we have done that, lets focus on the other Important features.
According to me, the next most important are HP and Speed.
Some would argue Sp. Attack and Sp. Defence but I think Ash is in for the long run, HP and Speed matter more here 

```python
# checking the distribution of HP
dataa = [go.Bar(
            x = nextstarter['HP'].values,
            
            width = 0.5,
            marker=dict(
            color = nextstarter['HP'].values,
            colorscale='Portland',
            showscale=True,
            reversescale = False
            ),
            opacity=0.6
        )]

layout= go.Layout(
    title= 'Distribution of HP',
    hovermode= 'closest',
    xaxis = dict(
        title = 'HP of Pokemon'
    ),
    showlegend= False
)
fig = go.Figure(data=dataa, layout=layout)
py.iplot(fig, filename='barplothp')
```

So choosing HP > 75 will remove half of our current Pokemon, lets do that 

```python
greathp = nextstarter['HP']>75
overall = nextstarter[greathp]
```

Another look at the dataset to see how the Type 1 feature is faring

```python
fg = sns.FacetGrid(data= overall, hue='Type 1', size = 8)
fg.map(plt.scatter, 'Attack','Defense').add_legend()
```

Cool all are there, now lets get on to Speed

```python
dataa = [go.Bar(
            x = overall['Speed'].values,
            
            width = 0.5,
            marker=dict(
            color = overall['Speed'].values,
            colorscale='Portland',
            showscale=True,
            reversescale = False
            ),
            opacity=0.6
        )]

layout= go.Layout(
    title= 'Distribution of Speed',
    hovermode= 'closest',
    xaxis = dict(
        title = 'Speed of Pokemon'
    ),
    showlegend= False
)
fig = go.Figure(data=dataa, layout=layout)
py.iplot(fig, filename='barplotspeed')
```

```python
# lets check for the mean, pretty sure 75 is good enough
np.mean(overall['Speed'].values)
```

```python
# cool so 75 it is 
greatspeed = overall['Speed']>75
overall = overall[greatspeed]
```

```python
# lets see how many Pokemon are left, if many are there then we will wean them out using Sp. Atk 
overall.shape
```

47 is good enough, lets check from which Generation our coveted 47 are from

```python
fg = sns.FacetGrid(data= overall, hue='Generation', size = 8)
fg.map(plt.scatter, 'Attack','Defense').add_legend()
```

Damn, none from 6th Gen, a shame 
and a huge number from 1st Gen, the Pokemon's are not made the same as they were back in the old days ;(
