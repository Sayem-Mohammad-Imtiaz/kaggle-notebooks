```python
input_cols=['longitude', 'latitude', 'housing_median_age', 'total_rooms',
       'total_bedrooms', 'population', 'households', 'median_income',
       'ocean_proximity']
input_df=housing_df[input_cols]
target_df=housing_df['median_house_value']
print(type(input_df))
print(type(target_df))
```

```python
from sklearn.model_selection import train_test_split
train_input,test_input,train_targets,test_targets=train_test_split(input_df,target_df,test_size=0.2,random_state=42)

print (train_input.shape, train_targets.shape)
print (test_input.shape, test_targets.shape)
```

```python
def estimate_charges(income, w, b):
    return w* income+b
```

```python
def rmse(targets, predictions):
    return np.sqrt(np.mean(np.square(targets - predictions)))
```

```python
w = 50
b = 100
```

```python
targets = train_targets
predicated = estimate_charges(train_input.median_income, w, b)
```

```python
rmse(targets, predicated)
```

```python
def try_parameters(w, b):
    income = housing_df.median_income
    target = housing_df.median_house_value
    predictions = estimate_charges(income, w, b)
    
    plt.plot(income, predictions, 'r', alpha=0.9);
    plt.scatter(income, target, s=8,alpha=0.8);
    plt.xlabel('income');
    plt.ylabel('House_value')
    plt.legend(['Prediction', 'Actual']);
    
    loss = rmse(target, predictions)
    print("RMSE Loss: ", loss)
```

```python
try_parameters(1000, 2000)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
inputs = train_input[['median_income']]
target = train_targets
print('inputs.shape :', inputs.shape)
print('targes.shape :', target.shape)
```

```python
model.fit(inputs, target)
```

```python
predictions = model.predict(inputs)
print(predictions)
```

```python
rmse(target, predictions)
```
