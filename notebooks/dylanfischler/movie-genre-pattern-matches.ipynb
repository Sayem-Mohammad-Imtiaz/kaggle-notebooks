```python
import pandas as pd
pd.set_option('mode.chained_assignment', None)

target_genre = "Thriller"
allowed_word_count = 2
name_pattern = "[tT]he [A-z]+ of [A-z ]+"
movies_df = pd.read_csv('/kaggle/input/imdb-extensive-dataset/IMDb movies.csv', low_memory=False)

genre_movies_df = movies_df[movies_df['genre'].str.contains(target_genre)]
```

```python
genre_movies_df.head()
```

```python
genre_size = len(genre_movies_df.index)
print("Total {} movies: {}".format(target_genre, len(genre_movies_df.index)))

majority_num = round(genre_size / 2)

genre_movies_df['word_count'] = genre_movies_df['title'].str.split().str.len()
genre_movies_df['over_allowed_word_count'] = genre_movies_df['word_count'] > allowed_word_count
genre_movies_df['matches_pattern'] = genre_movies_df['title'].str.contains(name_pattern)

genre_name_pattern_df = genre_movies_df[(genre_movies_df['matches_pattern'] == True) | (genre_movies_df['over_allowed_word_count'] == False)]

print("Min required for majority: {}".format(majority_num + 1))

pattern_match_count = len(genre_name_pattern_df.index)
print("Movies matching the pattern: {}".format(pattern_match_count))
print("Margin: {} ({}%)".format(pattern_match_count - majority_num, pattern_match_count/genre_size))
print("Pattern validity: {}".format(pattern_match_count > majority_num))
```
