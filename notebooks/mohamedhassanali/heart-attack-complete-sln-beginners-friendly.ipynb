# Modeling

```python
#creating dummies
```

```python
df_dummies = pd.get_dummies(df)
```

```python
#lets split our dataset
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test = train_test_split(df_dummies,test_size = 0.2 , random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
print('x_train : ' + str(x_train.shape)) 
print('x_test : ' + str(x_test.shape))
```

```python
x_features = [feature for feature in df.columns if feature not in  ['output','index']]
```

```python
x_features
```

```python
lg_r = LogisticRegression()
lg_r.fit(x_train[x_features] , x_train['output'])
```

```python
y_predicted = lg_r.predict(x_test[x_features])
```

```python
y_predicted
```

```python
from sklearn.metrics import accuracy_score

print('logistic regression accurracy is  : ' + str(round(accuracy_score(x_test['output'],y_predicted)*100 , 2))+ '%')
```

```python
rf_model = RandomForestClassifier()
rf_model.fit(x_train[x_features] , x_train['output'])
y_predicted = rf_model.predict(x_test[x_features])
print('RandomForest  accurracy is  : ' + str(round(accuracy_score(x_test['output'],y_predicted)*100 , 2))+'%')
```

```python
knn_model = KNeighborsClassifier()
knn_model.fit(x_train[x_features] , x_train['output'])
y_predicted = knn_model.predict(x_test[x_features])
print('KNN  accurracy is  : ' + str(round(accuracy_score(x_test['output'],y_predicted)*100 , 2)) + '%')
```

```python
gnb_model = GaussianNB()
gnb_model.fit(x_train[x_features] , x_train['output'])
y_predicted = gnb_model.predict(x_test[x_features])
print('Naive Bayes accurracy is  : ' + str(round(accuracy_score(x_test['output'],y_predicted)*100 , 2))+'%')
```

```python
#as you can see the logistic regression and naive_bayes gave the best predictions 
#the whole dataset is a good proof for me that not all models are bad .. I have aloways thought that Random forest is the best
# but as you can see its not the case for all cases 
```
