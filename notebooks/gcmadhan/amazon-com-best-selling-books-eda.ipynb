# **Amazon.com is one of the best online shopping site to buy books. in this notebook, lets do a Exploratory Data Analyis to review the top selling book from 2009 to 2019 and analyse the relations ship between the reviews, prices etc which helps the people to buy.**

**Reading the Data**

```python
df = pd.read_csv('../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
```

**Review the Fields and type of data**

There are 550 rows contains the best selling book details. 
There are 7 columns
1. Name - Name of the book
2. Author - Author name of the book
3. User Rating - Rating provided by the users ranging from 0 to 5
4. Reviews - no of reviews recevied for the books. inspite of postivie or negative reviews.
5. Year - Year in which the book is published
6. Genere - Gener of the book whether it is Fiction or Non-Fictions 

**let us see number of books distributed by year**

```python
eda_year=df.groupby('Year').sum().reset_index()
```

```python
eda_year
```

```python
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import matplotlib.pyplot as plt
```

Scatter plot to view the User Rating distributed by year
Based on the below chart, it is understood that best selling books have higher rating 

```python
df['Name_len']=df['Name'].str.len()
```

```python
px.scatter(df,x='Name_len',y='User Rating', color='Genre', size='User Rating', title="Relations ship between Name length and User Rating by Genre")
```

Relationshop between the year and Rating by Genre. where we can see that Non-Fiction has higher rating 

```python
px.bar(df,x='Year',y='User Rating',  color='Genre', barmode='group', title="User Rating by Year by Genre")
```

Every year Fiction has higer rating among best selling books than the Non-Fiction

```python
px.bar(df, x='Genre')
```

```python
df.head()
```

**Top 10 books with higher rating**

```python
df[['User Rating','Author','Name']].groupby(by=['Author','Name']).mean().sort_values(by=['User Rating'],ascending=False).head(10)
```

**Top Selling Authors**

```python
df[['Author','User Rating']].groupby(by='Author').mean().sort_values(by='User Rating',ascending=False).head(10)
```

**Top 5 selling books**

```python
df[['Name','User Rating','Year']].groupby(by='Name').mean().sort_values(by='User Rating',ascending=False).head(5)
```

```python
px.scatter_matrix(df,dimensions=['Price','User Rating'],color='Genre',size='Name_len')
```
