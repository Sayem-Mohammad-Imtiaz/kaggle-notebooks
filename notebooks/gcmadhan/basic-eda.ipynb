```python
df['DATE_OF_REGISTRATION'].dropna(axis=0,inplace=True)
```

```python
df['DATE_OF_REGISTRATION'].count()
```

```python
df['DATE_OF_REGISTRATION']=pd.to_datetime(df['DATE_OF_REGISTRATION'], errors = 'coerce', format='%d-%m-%Y')
```

```python
df['YEAR_REG']=pd.DatetimeIndex(df['DATE_OF_REGISTRATION']).year
```

```python
number_of_companies = df[['YEAR_REG','CORPORATE_IDENTIFICATION_NUMBER']]
```

```python
number_of_companies.dropna(axis=0,inplace=True)
```

```python
number_of_companies=number_of_companies.groupby(by='YEAR_REG').size().reset_index(name='No_of_companies')
```

```python
number_of_companies
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(30,10))
plt.xticks(rotation=45)
sns.barplot(data=number_of_companies,x='YEAR_REG',y='No_of_companies')
```

```python
no_of_companies_state = df[['YEAR_REG','CORPORATE_IDENTIFICATION_NUMBER','REGISTERED_STATE']]
```

```python
no_of_companies_state.dropna(axis=0,inplace=True)
```

```python
no_of_companies_state=no_of_companies_state.groupby(by=['YEAR_REG','REGISTERED_STATE']).size().reset_index(name='no_of_companies')
```

```python
no_of_companies_state
```

```python
plt.figure(figsize=(30,10))
plt.xticks(rotation=45)
sns.barplot(data=no_of_companies_state,x='YEAR_REG',y='no_of_companies',hue='REGISTERED_STATE')
```

```python
state='Tamil Nadu'
plot_data = no_of_companies_state[no_of_companies_state['REGISTERED_STATE']==state]
plt.figure(figsize=(30,10))
plt.xticks(rotation=45)
sns.barplot(data=plot_data,x='YEAR_REG',y='no_of_companies',hue='REGISTERED_STATE')
```
