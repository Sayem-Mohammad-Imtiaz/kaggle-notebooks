```python
from sklearn.preprocessing import LabelEncoder
```

```python
LB_Encode = LabelEncoder()
```

```python
data = df
```

```python
data.head()
```

```python
data['current_status']=LB_Encode.fit_transform(data['current_status'])
data['sex']=LB_Encode.fit_transform(data['sex'])
data['age_group']=LB_Encode.fit_transform(data['age_group'])
data['Race and ethnicity (combined)']=LB_Encode.fit_transform(data['Race and ethnicity (combined)'])
data['hosp_yn']=LB_Encode.fit_transform(data['hosp_yn'])
data['icu_yn']=LB_Encode.fit_transform(data['icu_yn'])
data['death_yn']=LB_Encode.fit_transform(data['death_yn'])
data['medcond_yn']=LB_Encode.fit_transform(data['medcond_yn'])
```

```python
data.drop(['cdc_report_dt'],axis=1,inplace=True)
```

```python
data.head()
```

```python
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
from sklearn.model_selection import train_test_split
```

```python
sns.heatmap(data.corr(), annot=True)
```

```python
data['death_yn'].value_counts()
```

```python
data['death_yn'].unique()
```

```python
##selecting only 50,000 records for each category for sampling
data1=pd.DataFrame()
for i in data['death_yn'].unique():
    data1=data1.append(data[data['death_yn']==i][:50000])
    
```

```python
len(data1)
```

```python
X = data1.drop(['death_yn'],axis=1)
y = data1['death_yn']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
model.fit(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, pred)
```

```python
print(classification_report(y_test,pred))
```

```python
sns.heatmap(confusion_matrix(y_test,pred), annot=True)
```
