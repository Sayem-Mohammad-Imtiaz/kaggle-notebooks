**Visualisation**

```python
colors = ['#75daef','#75daff','#f990f6','#75daef','#f990fc','#f990fc','#f990fc','#75daff']
dataset[['gender','age']].value_counts().plot.pie(explode=[0,0,0.2,0,0.2,0.2,0.2,0],autopct='%1.1f%%',colors=colors)
plt.ylabel('')
plt.title('Gender and Age Analysis')
```

Gender and age analysis shows that most of the target audiance is between the age group of 30-34 for both males and females.

```python
def sum_mean(data,col1,col2=[]):
    sum_df, mean_df = [] ,[]
    for i in data[col1].unique():
        d = data[data[col1] == i]
        s, m = [i], [i]
        if len(col2) == 0:
            col2 = [x for x in data.columns if x != col1]
        for j in col2:
            s.append(d[j].sum())
            m.append(d[j].sum())
        sum_df.append(s)
        mean_df.append(m)
    return sum_df,mean_df
```

```python
sum_df,mean_df = sum_mean(dataset,'age',['interest1','interest2','interest3'])
dt1 = pd.DataFrame(sum_df)
dt1.columns =['age','interest1','interest2','interest3']

dt1.plot.bar()
plt.xticks(np.arange(4),dt1['age'],rotation = 0)
plt.xlabel('Age')
plt.ylabel('Total Interest')
plt.title('Total Interest in Age groups')
```

Our main foucs should be for the age group 30-34, and their interest is 2 and 3

```python
sum_df,_ = sum_mean(dataset,'gender',['interest1','interest2','interest3'])
dt2 = pd.DataFrame(sum_df)
dt2.columns =['gender','interest1','interest2','interest3']
dt2.plot.bar()
plt.ylabel('Total Interests')
plt.xlabel('Gender')
plt.xticks(np.arange(2),dt2['gender'],rotation = 0)
plt.title('Interests in Gender')
```

Since their are more males watvhing the campaigns we can drive it towards their attention

```python
sum_df,_ = sum_mean(dataset,'gender',['impressions','spent'])
dt2 = pd.DataFrame(sum_df)
dt2.columns =['gender','impressions','spent']

plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.bar(dt2['gender'],dt2['impressions'],color=['#abcdef','#fedcba'])
plt.ylabel('Total Impressions')
plt.xlabel('Gender')
plt.xticks(np.arange(2),dt2['gender'],rotation = 0)

plt.subplot(1,2,2)
plt.bar(dt2['gender'],dt2['spent'],color=['#abcdef','#fedcba'])
plt.ylabel('Total Spents')
plt.xlabel('Gender')
plt.xticks(np.arange(2),dt2['gender'],rotation = 0)
plt.suptitle('Impressions and Spent in Gender')
```
