```python
df = df.drop(['Sunshine','Evaporation','Cloud3pm','Cloud9am','Date', 'Location'], axis=1)
df.head()
```

```python
# dealing with categorical variable
df['RainToday'] = df['RainToday'].replace({'Yes' : 1 , 'No' : 0})
df['RainTomorrow'] = df['RainTomorrow'].replace({'Yes' : 1 , 'No' : 0})

cat_variable = ['WindGustDir','WindDir9am','WindDir3pm']
df = pd.get_dummies(df,columns=cat_variable,drop_first=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
y=df['RainTomorrow']
X = df.drop(['RainTomorrow'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=32, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

Using Gradient Boosting model

```python
from sklearn.ensemble import GradientBoostingClassifier
gbmclr = GradientBoostingClassifier()
```

```python
gbmclr.fit(X_train,y_train)
ygbm_pred = gbmclr.predict(X_test)
```

```python
accuracy_score(y_test,ygbm_pred)
```

```python
confusion_matrix(y_test,ygbm_pred)
```
