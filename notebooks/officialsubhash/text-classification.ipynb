```python
import re
from string import printable
import matplotlib.pyplot as plt

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, plot_precision_recall_curve, plot_confusion_matrix
from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB, GaussianNB
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from xgboost import XGBClassifier, XGBRFClassifier

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("../input/amazon-music-reviews/Musical_instruments_reviews.csv", encoding='latin')
df.head()
```

```python
df.shape
```

```python
columns = ['reviewText', 'overall']
df = df[columns]
df.head()
```

```python
df['target'] = df.overall.apply(lambda x: 1 if x>=4.0 else 0)
df.head()
```

```python
df.target.value_counts()
```

```python
df.dropna(inplace=True)
```

```python
STOPWORDS = stopwords.words('english') + list(printable)
lemma = WordNetLemmatizer()

def text_preprocess(text):
    text = text.lower()
    tokens = [lemma.lemmatize(word) for word in word_tokenize(text) if word not in STOPWORDS]
    text = ' '.join(tokens)
    
    return text
```

```python
df['clean_text'] = df.reviewText.apply(text_preprocess)
df.head()
```

```python
tfidf = TfidfVectorizer(max_features=500)
data = tfidf.fit_transform(df.clean_text)
```

## Classification

```python
X, y = data, df.target
```

```python
### Logistic Regression

log_reg = LogisticRegression().fit(X,y)
plot_confusion_matrix(log_reg,X,y, cmap='Blues');
```

```python
## Naive bayes - Gaussian

gb_clf = GaussianNB().fit(X.toarray(),y)
plot_confusion_matrix(gb_clf,X.toarray(),y, cmap='Blues');
```

```python
## Naive Bayes Bernouli
bnb = BernoulliNB().fit(X.toarray(), y)
plot_confusion_matrix(bnb,X.toarray(),y, cmap='Blues');
```

```python
## Random Forest

rndf = RandomForestClassifier().fit(X, y)
plot_confusion_matrix(rndf,X,y, cmap='Blues');
```
