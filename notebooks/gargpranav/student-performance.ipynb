```python
import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
## Null values
df.isna().any()
```

No Null Values in the Dataset

```python
## Categorical Values unique values
print('Gender Unique Values',df.gender.nunique())
print('race Unique Values',df['race/ethnicity'].nunique())
print('parental level of education Unique Values',df['parental level of education'].nunique())
print('lunch Unique Values',df['lunch'].nunique())
print('test preparation course Unique Values',df['test preparation course'].nunique())
```

```python
df.gender = np.where(df.gender == 'female', 0, 1)
```

```python
df['race'] = df['race/ethnicity']
df['parental_education'] = df['parental level of education']
df['test_preparation'] = df['test preparation course']
df.drop(['race/ethnicity', 'parental level of education', 'test preparation course'], axis=1, inplace=True)
```

```python
df.head()
```

```python
df['total_score'] = df['math score'] + df['reading score'] + df['writing score']
```

```python
print("Avg Math Score:- ", df['math score'].mean())
print("Avg writing Score:- ", df['writing score'].mean())
print("Avg reading Score:- ", df['reading score'].mean())
print("Avg total Score:- ", df['total_score'].mean())
```

```python
## categorical values graph

category_columns = ['race', 'parental_education', 'test_preparation']
for column in category_columns:
    plt.figure(figsize=(10,4))
    sns.lineplot(x = df[column], y=df.total_score, hue = df.gender, estimator=np.mean)
    plt.show()
```

As we can see that 
1. Group E has highest total score among all groups
2. Parents who has bachelor's or master's Degree, their kids are doing well then other kids
3. One more thing to notice that Female kids are performing well than male kids

```python
plt.figure(figsize=(12,6))
plt.title('PARENTS LEVEL OF EDUCATION')
sns.countplot(x = df.parental_education)
```
