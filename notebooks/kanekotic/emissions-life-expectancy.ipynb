We want to understand if there is a correlation between the emission of CO2 and the life expectancy in different countries.

```python
import pandas
import numpy as np
```

```python
emission_data = pandas.read_csv('../input/co2-and-ghg-emission-data/emission data.csv')
emission_data.head()
```

```python
life_expectancy_data = pandas.read_csv('../input/life-expectancy-who/Life Expectancy Data.csv')
life_expectancy_data.head()
```

```python
#What is the average live expectancy globally?
expected_data=life_expectancy_data[life_expectancy_data.Year >= 2010]['Life expectancy ']
expected_data.dropna().apply(np.int64)
print(f'Global life expectancy from 2010 to 2015 {expected_data.dropna().apply(np.int64).mean()}')
```

```python
#What is the difference of life expectancies in between Developed and Developing Countries?
expected_data=life_expectancy_data[life_expectancy_data.Year >= 2010][['Status','Life expectancy ']].dropna()
selection = expected_data.groupby('Status').agg(life_expectancy =('Life expectancy ', np.mean))
selection.head(10)
```

```python
#What are the top 10 countries with lower life expectancy?
expected_data=life_expectancy_data[life_expectancy_data.Year >= 2010][['Country','Life expectancy ']].dropna()
selection = expected_data.groupby('Country').agg(life_expectancy =('Life expectancy ', np.mean))
selection.sort_values(by='life_expectancy').head(10)
```

```python
#What are the top 10 countries with higher life expectancy?
expected_data=life_expectancy_data[life_expectancy_data.Year >= 2010][['Country','Life expectancy ']].dropna()
selection = expected_data.groupby('Country').agg(life_expectancy =('Life expectancy ', np.mean))
selection.sort_values(by='life_expectancy', ascending=False).head(10)
```
