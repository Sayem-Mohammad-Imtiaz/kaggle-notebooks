**PLOTANDO OS GRÁFICOS**

```python
#Comparando os suicidios com o PIB per capita e agrupando pelas gerações
plt.figure(figsize=(10,9))
sns.pairplot(x_vars='gdp_per_capita',y_vars='suicides',data=df, height=10,hue='generation')
```

Aparentemente, em lugares pobres, há muito suicídio. À medida que a renda aumenta, o suicídio diminui juntos. No entanto, a partir de um ponto (20k), o suicídio tende a aumentar novamente.

```python
plt.figure(figsize=(10,9))
sns.barplot(x = 'age', y='suicides', hue='sex', data=df.groupby(["age","sex"]).sum().reset_index()).set_title('Suicidios por faixa etária')
plt.ylabel('SUICÍDIOS')
plt.xlabel('IDADE')
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x = 'age', y='suicidesper100kpop', hue='sex', data=df.groupby(["age","sex"]).sum().reset_index()).set_title('Suicidios por 100.000 pessoas')
plt.ylabel('SUICÍDIOS')
plt.xlabel('IDADE')
plt.xticks(rotation=90)
```

Comparando os dois gráfico a cima obeserva-se que no primeiro gráfico, usando a variavel **Suicides** a taxa de suicídio e maior nas pessoas com idade entre 35-54 anos, no segundo gráfico usando a variavel **Suicidesper100kpop** a taxa de suicídio é maior para as pessoas com mais de 75 anos. 

```python
plt.figure(figsize=(10,9))
sns.barplot(x='sex', y='suicides', hue='generation', data=df)
```

O gráfico acima mostra o número de suicídios cometidos por pessoas de diferentes gerações em ambos os sexos. Podemos ver que o número de suicídios é maior na geração dos Boomers (1946 - 1964), enquanto que é menor na geração Z (2001 - 2010)

```python
suic_sum = pd.DataFrame(df['suicides'].groupby(df['country']).sum())
suic_sum = suic_sum.reset_index().sort_index(by='suicides',ascending=False)
most_cont = suic_sum.head(5)
fig = plt.figure(figsize=(10,10))
plt.title('Os 5 países que mais deve casos de suicídios.')
sns.set(font_scale=2)
sns.barplot(y='suicides',x='country',data=most_cont,palette="Blues_d")
plt.xticks(rotation=45)
plt.ylabel('Suicídios');
plt.xlabel('Países')
plt.tight_layout()
```

```python
plt.figure(figsize=(10,10))
plt.title('Gráfico da População por ano')
sns.lineplot(x='year',y='population',hue='sex',data=df)
```

A população de mulheres e maior que a de homens, ainda assim a quantidade de homens com suicídio e maior que a das mulheres

```python
#Gráfico de correlação das variaveis 
f, ax = plt.subplots(figsize=(10, 8))
sns.heatmap(df.corr(), vmin=-1,vmax=1,cmap='coolwarm', annot=True,fmt='.2f', linewidths=.5, ax=ax)
```

As maiores correlações são entre população e PIB, uma vez que os países ricos, em geral, são mais populosos. Além disso, entre o número de suicídios e a população, se houver mais pessoas, mais suicídios.
