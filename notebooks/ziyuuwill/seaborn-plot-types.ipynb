```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
vgsales = pd.read_csv('../input/videogamesales/vgsales.csv')
```

```python
print(vgsales.shape)
vgsales.head()
```

```python
vgsales.dropna(inplace=True)
```

```python
vgsales['year'] = vgsales['Year'].astype(int)
```

```python
pd.DataFrame(vgsales.groupby('Platform')['JP_Sales'].sum()).reset_index().sort_values('JP_Sales', ascending=False).iloc[:5]
```

```python
ps_sales = vgsales[vgsales['Platform'] == 'PS']
```

```python
ps_sales_year = pd.DataFrame(ps_sales.groupby('year')[['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']].sum()).reset_index()
ps_sales_year
```

# Trends
A trend is defined as a pattern of change.

## sns.lineplot
**Line charts** are best to show trends over a period of time, and multiple lines can be used to show trends in more than one group.

```python
plt.figure(figsize=(14,6))

plt.title("PS")

sales_cols = ps_sales_year.drop(['year','Global_Sales'], axis=1).columns

for col in list(sales_cols):
    sns.lineplot(data=ps_sales_year, x=ps_sales_year['year'], y=ps_sales_year[col], label=col)

plt.xlabel("Year")
```

# Relationship
There are many different chart types that you can use to understand relationships between variables in your data.

## sns.barplot
**Bar charts** are useful for comparing quantities corresponding to different groups.

```python
plt.figure(figsize=(10,6))
plt.title("PS Global Sales")

sns.barplot(data=ps_sales_year, x=ps_sales_year['year'], y=ps_sales_year['Global_Sales'])
plt.ylabel("Global_Sales")
```

## sns.heatmap
**Heatmaps** can be used to find color-coded patterns in tables of numbers.

```python
plt.figure(figsize=(14,7))
plt.title('PS Sales')

sns.heatmap(data=ps_sales_year.set_index('year').drop('Global_Sales', axis=1), annot=True)
```
