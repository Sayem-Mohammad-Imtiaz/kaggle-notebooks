## Missing Values

```python
faltantes_percentual = ((df.isnull().sum()) / len(df["class"]) * 100).round(2)
print(f"Tabela faltantes_percentual: \n{faltantes_percentual}\n")
```

### Using one-hot encoding

```python
y = df["class"]
X = df.drop("class", axis = 1)
```

```python
X = pd.get_dummies(X)
```

```python
y = y.replace("e", 0)
y = y.replace("p", 1)
```

## Extra Trees aplication

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import StratifiedKFold, cross_val_score

model = ExtraTreesClassifier(n_jobs = -1)
skf = StratifiedKFold(n_splits = 5, shuffle = True)
resultado = cross_val_score(model, X,y, cv = skf)
print(resultado.mean().round(4))
```
