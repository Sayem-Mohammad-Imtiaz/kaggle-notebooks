```python
#Nested Functions

#Example: name and surname combining method and printer method
#Nested Method Starting.
def printerFunc(name,surname): #External method
    def comibinedNames(name,surname):#internal method
        return name+" "+surname
    print(name,surname)
#Nested Method Finished.
name="Hakan"
surname="Yusufoglu"
printerFunc(name,surname)
```

```python
#Default Arguments(parameter)
#Example mean method
def averageFunc(score1,score2,score3=48): #score3= Default value so default argumen
    sumValues=score1+score2+score3
    return sumValues/3
print(averageFunc(10,10)) #score3 value=48 because default argumen=48
#and
print(averageFunc(10,10,10)) #score3 value=10 because changed with 10
```

```python
#Flexible Arguments
#Example many value print

def sumFunc(*values):
    for i in values:
        print(i)
sumFunc(10) #1 parameter
print("")
sumFunc(20,30)#two parameter etc (many parameter)
```

```python
#Lamda Functions
#Briefly faster function
#Example mean method

avarage=lambda score1,score2,score3=48: (score1+score2+score3)/3 #up: normal method
print(avarage(10,10))
print(avarage(10,10,10))
```

**LIST COMPREHENSİON**
We use list comprehension for data analysis often.

```python
#Example: Happiness Score mean
def meanFunction_2():
    sumValuable=0
    counter=0
    meanValuable=0
    for i in database["Happiness_Score"]:
        counter+=1
        sumValuable+=i
    meanValuable=sumValuable/counter
    return meanValuable
print(meanFunction_2())
```

```python
database.head(10)
meanScore=sum(database["Happiness_Score"])/len(database["Happiness_Score"])
#Example happy or unhappy countries with dataframe(World Happiness Report)
database["Happy_Status"]=["Happy" if i>meanScore else "UnHappy" for i in database.Happiness_Score]
database.loc[60:75,["Country","Happiness_Score","Happy_Status"]]
```

**PIVOTING DATA FRAMES**
