# FEATURE SCALLING :
it is a technique to standardize the independent variable/features present in the data in fixed range

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../input/social-network-ads/Social_Network_Ads.csv')
```

```python
df.head()
```

```python
df=df.iloc[:,2:5]
```

```python
df.head()
```

**sample() is used to see random rows from the data**

```python
df.sample(5)
```

```python
X=df.drop('Purchased',axis=1)
y=df['Purchased']
```

```python
X.head(1)
```

```python
y.head(1)
```

# Train test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

# Standard Scaler:
fit--> mean learn --> only training dataset
transform-->on the both train and test

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
scaler.mean_
```

```python
X_train_scaled=pd.DataFrame(data=X_train_scaled,columns=X_train.columns)
X_test_scaled=pd.DataFrame(data=X_test_scaled,columns=X_test.columns)
```

```python
X_train_scaled
```

```python
np.round(X_train.describe(),1)
```

```python
np.round(X_train_scaled.describe(),1)
```

# EFFECT OF SCALLING ON THE DATASET FOCUS ON THE GRAPH

```python
fig, (s1,s2)=plt.subplots(ncols=2, figsize=(12,5))
s1.scatter(X_train['Age'],X_train['EstimatedSalary'])
s1.set_title('Befor scalling')
s2.scatter(X_train_scaled['Age'],X_train_scaled['EstimatedSalary'],color='red')
s2.set_title('After scalling')
plt.show()
```

```python
fig, (ax1,ax2)=plt.subplots(ncols=2, figsize=(12,5))
ax1.set_title('Befor scalling')
sns.kdeplot(X_train['Age'],ax=ax1)
sns.kdeplot(X_test['EstimatedSalary'],ax=ax1)

ax2.set_title('After scalling')
sns.kdeplot(X_train_scaled['Age'],ax=ax2)
sns.kdeplot(X_test_scaled['EstimatedSalary'],ax=ax2)
plt.show()
```

```python
fig, (ax1,ax2)=plt.subplots(ncols=2, figsize=(12,5))
ax1.set_title('Befor scalling age distrubution')
sns.kdeplot(X_train['Age'],ax=ax1)
ax2.set_title('After scalling age distrubution')
sns.kdeplot(X_train_scaled['Age'],ax=ax2)
plt.show()
```

```python
fig, (ax1,ax2)=plt.subplots(ncols=2, figsize=(12,5))
ax1.set_title('Befor scalling salary distrubution')
sns.kdeplot(X_train['EstimatedSalary'],ax=ax1)
ax2.set_title('After scalling salary distrubution')
sns.kdeplot(X_train_scaled['EstimatedSalary'],ax=ax2)
plt.show()
```

# Why scalling is important so you see distinct through this algorithm

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr_scalled=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
lr_scalled.fit(X_train_scaled,y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred_scaled=lr_scalled.predict(X_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("actual",accuracy_score(y_pred,y_test))
print("Scaled",accuracy_score(y_pred_scaled,y_test))
```
