```python
df = df.drop(['Unnamed: 0'], axis = 1)
df.shape
df.columns
```

> # ***Heat Map Showing relationship***

```python
plt.figure(figsize = (15,10))

corr =df.corr()
sns.heatmap(corr, cmap="coolwarm", linewidths=1)
```

```python
Q1 = df.quantile(0.2)
Q3 = df.quantile(0.8)
IQR = Q3 - Q1
#print(IQR)
```

# > > ***Removing Outliers***

```python
df_out = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df_out.head()
                                  
```

# Pairplot after removing outliers

```python
sns.pairplot(df_out, vars=['carat', 'depth','table','price', 'x','y','z'])
```

> # ***Using Goupby Function***

```python
# multiple groupby (carat and x both)
df1 = df_out.groupby(['x', 'carat']).count()['price']
plt.figure(figsize=(10,7))  
# plot the result
df1.plot()
plt.xticks(rotation=45)
plt.show()
```

```python
# multiple groupby (carat and depth both)
df1 = df_out.groupby(['carat', 'price']).count()['depth']
plt.figure(figsize=(10,7))  
# plot the result
df1.plot()
plt.xticks(rotation=45)
plt.show()
```
