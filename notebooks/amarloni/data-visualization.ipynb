> ***Feature Engineering
 I have split the Start Time in different features like month, year, time, week day etc as example**** 

```python
import datetime
df['Start_Time']= pd.to_datetime(df['Start_Time'])
df['hour']= df['Start_Time'].dt.hour
df['year']= df['Start_Time'].dt.year
df['month']= df['Start_Time'].dt.month
df['week']= df['Start_Time'].dt.week
df['day']= df['Start_Time'].dt.weekday_name
df['quarter']= df['Start_Time'].dt.quarter
df['time_zone']= df['Start_Time'].dt.tz
df['time']= df['Start_Time'].dt.time
```

```python
df.head()
```

```python
plt.figure(figsize =(10,5))
df.groupby(['year']).size().sort_values(ascending=True).plot.bar()
```

> ***How are the accident rates over the year? Increaseing........***

```python
plt.figure(figsize =(15,5))
df.groupby(['month']).size().plot.bar()
```

> ***Above you can see the month wise accident stats***

```python
plt.figure(figsize =(15,5))
df.groupby(['year', 'month']).size().plot.bar()
plt.title('Number of accidents/year')
plt.ylabel('number of accidents')
```

> ***Above you can see the year  & month wise accident stats***

```python
df.groupby(['day']).size().plot.pie(figsize=(10,10))
```

***Above you can see that on Saturday & Sundays the accidents are minimum***

```python
plt.figure(figsize =(10,5))
df.groupby(['hour']).size().plot.bar()
plt.title('At which hour of day accidents happen')
plt.ylabel('count of accidents')
```

***Above Graph Shows at which time of the day the accidents are maximum, early morning during the office going time***

```python
df['day_zone'] = pd.cut((df['hour']),bins=(0,6,12,18,24), labels=["night", "morning", "afternoon", "evening"])
plt.figure(figsize =(10,5))
df.groupby(['day_zone']).size().plot.bar()
```

> ***Above I have used Pandas cut function to split the time in 4 zones of morning evening afternoon & night this shows during morning & afernoon the accidents are at the max.***
