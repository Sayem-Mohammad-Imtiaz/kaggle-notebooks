```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

importing required libraries

```python
df=pd.read_csv("../input/movies-on-netflix-prime-video-hulu-and-disney/MoviesOnStreamingPlatforms_updated.csv")
```

reading the file into a variable df

```python
df.shape
```

shape attribute gives the number of rows and columns of the dataset

```python
df.describe(include='all')
```

describe is used to obtain details like mean,percentile,freq etc.

```python
df.isnull().sum()
```

isnull().sum()returns the number of missing values in the data set

```python
plt.subplot(221)

df['Prime Video'].value_counts().plot(kind='bar', title='Amazon Prime Video', figsize=(16,9))

plt.xticks(rotation=0)

plt.subplot(222)

df['Disney+'].value_counts().plot(kind='bar', title='Disney+')

plt.xticks(rotation=0)

plt.subplot(223)

df['Hulu'].value_counts().plot(kind='bar', title='Hulu')

plt.xticks(rotation=0)

plt.subplot(224)

df['Netflix'].value_counts().plot(kind='bar', title='Netflix')

plt.xticks(rotation=0)

plt.show()
```

Observation:here we have drawn bar graph for values count for different OTT platforms.In this graph 0 on x-axis represents non-null values and 1 represents null values present.By this graph we can conclude that Amazon prime video has highest null value count and Hulu has least null value count

```python
df.boxplot()
```

Observation: We can see that Unnamed and Id have similar boxplots with a minimum at 0 and maximum above 8k, median around 5k.There are outliers present at Year, Hulu, Disney plus and Run time

boxplot displays the distribution of data on basis of minimum,maximum,median,first quartile and second quartile.it also indicates whether the data is symmetrical are not.

```python
sns.distplot(df['Runtime'])
```

observation: run time between 0 to 100 has highest density

distplot gives as histogram with a line over it.distplot can be used to represent the univariate distribution of data.

```python
corr = df.corr()
sns.heatmap(corr, annot=True, square=True)
plt.yticks(rotation=2)
plt.show()
```

annot = True makes sure that the values of each cell appear on the chart and yticks sets the y-axis values

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(df["Hulu"], df["Netflix"])
ax.set_xlabel("Hulu")
ax.set_ylabel("Netflix")
plt.show()
```

The subplots() function takes three arguments that describes the layout of the figure. The layout is organized in rows and columns

```python
df.hist(figsize=(10, 12), bins=50, xlabelsize=8, ylabelsize=8); 
```

this plots a histogram with width as 10 and height as 12.bins represents the bars of histogram.bins can be changed as per our wish

```python
sns.factorplot("Age",data=df,kind="count")
```

Observation: The age of 18+  have highest count

factorplot is used to represent different types of categorical plot.the type of plot can be specified using 'kind' parameter
