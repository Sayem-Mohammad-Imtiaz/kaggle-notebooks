## Let's find out how accurate is this API?

```python
df = pd.read_csv('/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv')
df.head()
```

```python
df['sentiment'].replace(to_replace ="positive", value ="1", inplace=True)
df['sentiment'].replace(to_replace ="negative", value ="0", inplace=True)
```

```python
df.head(3)
```

```python
reviews_sample = df['review'][:100]
reviews_sample[:5]
```

```python
import re
def sentiment140(reviews):
    preds = []
    for review in reviews:
        values = {'data': [{'text': review}]} 

        data = json.dumps(values)
        response = urllib.request.urlopen(url, data=data.encode("utf-8"))
        page = response.read()
        
        page = page.decode("utf-8")
        polarity = re.search('polarity":.', str(page)).group(0)[-1]
        preds.append(polarity)
    return preds
```

```python
predictions = sentiment140(reviews_sample)
```

```python
predictions = np.array(predictions, dtype=np.int16)
predictions[predictions == 4] = 1
predictions[:10]
```

```python
labels = (df['sentiment'][:100]).values
labels = np.array(labels, dtype=np.int16)
```

```python
# comparing results
labels[:10]
```

```python
print('Accuracy is: ',np.sum(np.equal(labels, predictions)) / len(labels))
```
