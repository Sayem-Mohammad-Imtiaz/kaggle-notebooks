### Scope
A variable is only available from inside the region it is created. This is called scope.
* global: a variable created in the main body of the Python code is a global variable
* local: a variable created inside a function belongs to the local scope of that function, and can only be used inside that function.
* built in scope: names in predefined built in scope module such as print, len

```python
# guess prints what
x = 4
def f():
    x = 2
    return x
print(x)      # x = 4 global scope
print(f())    # x = 2 local scope
```

```python
# What if there is no local scope
x = 7
def f():
    y = 2*x        # there is no local scope x
    return y
print(f())         # it uses global scope x  
# First local scope searched, then global scope searched, if two of them cannot be found lastly built in scope searched.
```

### NESTED FUNCTION
* A function defined inside another function is called a nested function. 
* There is a LEGB rule that is search local scope, enclosing function, global and built in scopes, respectively.

```python
# Nested function
def square():
    """ return square of value """
    def add():
        """ add two local variable """
        x = 4
        y = 8
        z = x + y
        return z
    return add()**2
print(square())    
```

### DEFAULT and FLEXIBLE ARGUMENTS
* Default argument example:
<br> def f(a, b=1):
        """ b = 1 is default argument"""
* Flexible argument example:
<br> def f(*args):
       """ *args can be one or more"""
<br>def f(** kwargs)
       """ **kwargs is a dictionary"""
       
<br><br> Example for this arguments:

```python
# Default arguments
def f(a, b = 1, c = 10):
    y = a + b + c
    return y
print(f(5))
# what if we want to change default arguments
print(f(5,4,3))
```

```python
# flexible arguments *args
def f(*args):
    for i in args:
        print(i)
f(1)
print("")

f(1,2,3,4)

# flexible arguments **kwargs that is dictionary
def f(**kwargs):
    """ print key and value of dictionary"""
    for key, value in kwargs.items():           
        print(key, " ", value)
f(country = 'turkey', capital = 'ankara', population = 5504000)
```

### List Comprehension 

List comprehension is a method for list building and is a method to create lists effectively and with little code

```python
# Example of list comprehension


num1 = [3,2,1]
num2 = [i +1 for i in num1]
num3 = [i * 4 for i in num1]

print(num2)
print(num3)
```

**[i + 1 for i in num1 ]**: list of comprehension
<br> **i +1**: list comprehension syntax
<br> **for i in num1**: for loop syntax
<br> **i**: iterator
<br> **num1**: iterable object

```python
# Conditionals on iterable 

num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)
```

```python
# Another Example

num1 = [6, 45,10,15, 34,23]
num2 = [i**2 if i == 34 else i-5 if i < 15 else i+5 for i in num1]
print(num2)

num3 = [i*i if i >20 else i+ 5 if i > 10 else i-5 for i in num1]
print(num3)
```

### Explotary Data Analysis

<br>What is quartile?

In general, the goal of quantiles is to divide a data set into sections.
Dividing the sample into four equal parts thus creates three quartiles, which are also known as the upper quartile, median and lower quartile.

* First quantile: 25%
* Median / Second quantile: 50%
* Third quantile: 75%
* Outliers: is a statistical observation that deviates significantly from other members of a sample.
* std: standard deviation
* mean: average of entries
* count: number of entries
<br>Example:

<br>Sequenz: 5,9,8,3,0,8,2,7,12 => 0,2,3,5,7,8,8,9,12

<br>Lower quartile: 
* is the median in between the smallest number and the median i.e. in between 0 and 7, which is 3.

<br>Median: 
* sort the data in ascending order and select the record in the middle
* for even number of data, the average of the two data in the middle is calculated
* It would be 7 

<br>Upper quartile:
* you find the median between the median and the largest number i.e. between 7 and 12, which will be 8 according to the question above.
