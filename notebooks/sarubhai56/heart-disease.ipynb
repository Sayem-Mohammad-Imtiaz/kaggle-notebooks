```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import re

train = pd.read_csv("../input/heart.csv")
train.dropna()
```

```python

corr = train.corr()
corr.style.background_gradient()
```

```python
  from sklearn import svm
  from sklearn.model_selection import train_test_split
  
  train1 = train.copy()
  
  feature_df = train1[["age","sex","cp","trestbps","thalach","chol","restecg","exang","oldpeak","slope","ca","thal"]]
  x = np.asarray(feature_df)
  y = np.asarray(train["target"].astype('int'))
  
  x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 5)
  
  clf = svm.SVC(kernel='linear')
  clf.fit(x_train,y_train)
  yhat = clf.predict(x_test)
from sklearn.metrics import jaccard_similarity_score
print(print(jaccard_similarity_score(y_test,yhat)))
```

```python
############### MODEL VIA LOGISTIC REGRESSION

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(C=0.01,solver='liblinear').fit(x_train,y_train)

y_hat = lr.predict(x_test)

from sklearn.metrics import jaccard_similarity_score
print(print(jaccard_similarity_score(y_test,yhat)))
```

```python
#################### SVM LINEAR

from sklearn import svm
from sklearn.model_selection import train_test_split
  
train1 = train.copy()
  
feature_df = train1[["age","sex","cp","trestbps","exp_thalach","log_chol","restecg","exang","oldpeak","slope","ca","thal"]]
x = np.asarray(feature_df)
y = np.asarray(train["target"].astype('int'))
  
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 5)
  
clf = svm.SVC(kernel='linear')
clf.fit(x_train,y_train)
yhat = clf.predict(x_test)
from sklearn.metrics import jaccard_similarity_score
print(jaccard_similarity_score(y_test,yhat))
```

```python
############     using random forest model

# Feature Scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()  
x_train = sc.fit_transform(x_train)  
x_test = sc.transform(x_test) 
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=20, random_state=7)  
regressor.fit(x_train, y_train)  
y_pred = regressor.predict(x_test)

for i in range(len(y_pred)):
  if y_pred[i]>=0.5:
    y_pred[i]=1
  else:
    y_pred[i]=0
#print(y_pred)


#prediction_int = y_pred[:,1] >= 0.3 # if prediction is greater than or equal to 0.3 than 1 else 0
#prediction_int = prediction_int.astype(np.int)


from sklearn.metrics import jaccard_similarity_score
print(jaccard_similarity_score(y_test,y_pred))
```
