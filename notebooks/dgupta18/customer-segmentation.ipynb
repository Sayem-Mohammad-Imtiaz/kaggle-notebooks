```python
df = pd.read_csv('/kaggle/input/customer-segmentation-tutorial-in-python/Mall_Customers.csv',index_col = 'CustomerID')
df
```

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
df['Gender'] = label.fit_transform(df['Gender'])
```

```python
exp.scatter_3d(data_frame=df,x = 'Age',y = 'Annual Income (k$)',z = 'Spending Score (1-100)')
```

```python
sns.countplot(x = 'Gender',data = df)
```

```python
sns.pairplot(df)
```

```python
ssd = []
K = range(1,10)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(df)
    ssd.append(km.inertia_) 
```

```python
plt.figure(figsize=(10,6))
plt.plot(K, ssd, 'bx-')
plt.xlabel('k')
plt.ylabel('ssd')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmean = KMeans(n_clusters=5)
kmean.fit(df)
```

```python
kmean.cluster_centers_
```

```python
pred = kmean.labels_
```

```python
exp.scatter(data_frame=df,x = 'Annual Income (k$)',y = 'Spending Score (1-100)',color = pred)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_model = pca.fit_transform(df)
data_transform = pd.DataFrame(data = pca_model, columns = ['PCA1', 'PCA2'])
data_transform['Cluster'] = pred
```

```python
plt.figure(figsize=(8,8))
g = sns.scatterplot(data=data_transform, x='PCA1', y='PCA2',palette='gnuplot', hue='Cluster')
title = plt.title('Countries Clusters with PCA')
```
