```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
from scipy import stats
import plotly.express as ex
import plotly.graph_objs as go
import plotly.offline as pyo
pyo.init_notebook_mode()


plt.rc('figure',figsize=(17,12))
```

```python
states = ["AL - Alabama", "AK - Alaska", "AZ - Arizona", "AR - Arkansas", "CA - California", "CO - Colorado",
"CT - Connecticut", "DC - Washington DC", "DE - Deleware", "FL - Florida", "GA - Georgia",
"HI - Hawaii", "ID - Idaho", "IL - Illinois", "IN - Indiana", "IA - Iowa",
"KS - Kansas", "KY - Kentucky", "LA - Louisiana", "ME - Maine", "MD - Maryland",
"MA - Massachusetts", "MI - Michigan", "MN - Minnesota", "MS - Mississippi",
"MO - Missouri", "MT - Montana", "NE - Nebraska", "NV - Nevada", "NH - New Hampshire",
"NJ - New Jersey", "NM - New Mexico", "NY - New York", "NC - North Carolina",
"ND - North Dakota", "OH - Ohio", "OK - Oklahoma", "OR - Oregon", "PA - Pennsylvania",
"RI - Rhode Island", "SC - South Carolina", "SD - South Dakota", "TN - Tennessee",
"TX - Texas", "UT - Utah", "VT - Vermont", "VA - Virginia", "WA - Washington", "WV - West Virginia",
"WI - Wisconsin", "WY - Wyoming","DE - Delaware","DC - District of Columbia"]
states = {s.split(' - ')[1]:s.split(' - ')[0] for s in states}
```

<h1 style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; font-weight: normal; text-align: center;"><span data-preserver-spaces="true" style='color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; font-size: 24px; font-family: "Times New Roman", Times, serif;'>The above dictionary will help us transform all the state names in our dataset into a uniform format which the &quot;plotly&quot; library accepts as part of the choropleth method.</span></h1>

```python
e_data = pd.read_csv('/kaggle/input/2020-us-general-election-turnout-rates/2020 November General Election - Turnout Rates.csv')
e_data.head(3)
```

<a id="1.1"></a>
<h3 style="background-color:skyblue;font-family:newtimeroman;font-size:200%;text-align:center">Data Preprocessing</h3>


```python
e_data['VEP Turnout Rate'] = e_data['VEP Turnout Rate'].apply(lambda x: float(x.replace('%','')))
e_data['% Non-citizen'] = e_data['% Non-citizen'].apply(lambda x: float(x.replace('%','')))
e_data.loc[e_data['Vote for Highest Office (President)'].notna().index,'Vote for Highest Office (President)'] = e_data[e_data['Vote for Highest Office (President)'].notna()]['Vote for Highest Office (President)'].apply(lambda x: float(x.replace(',','')))


e_data['Voting-Eligible Population (VEP)'] = e_data['Voting-Eligible Population (VEP)'].apply(lambda x: float(x.replace(',','')))
e_data['Prison'] = e_data['Prison'].apply(lambda x: float(x.replace(',','')))
e_data['Probation'] = e_data['Probation'].apply(lambda x: float(x.replace(',','')))
e_data['Parole'] = e_data['Parole'].apply(lambda x: float(x.replace(',','')))
e_data['Total Ineligible Felon'] = e_data['Total Ineligible Felon'].apply(lambda x: float(x.replace(',','')))
e_data.loc[e_data['Overseas Eligible'].notna().index,'Overseas Eligible'] = e_data[e_data['Overseas Eligible'].notna()]['Overseas Eligible'].apply(lambda x: float(x.replace(',','')))
e_data['Total Ballots Counted (Estimate)'] = e_data['Total Ballots Counted (Estimate)'].apply(lambda x: float(x.replace(',','')))


e_data.loc[1:,'State'] = e_data.loc[1:,'State'].apply(lambda x: states[x]).copy()
```

<h3 style="background-color:skyblue;font-family:newtimeroman;font-size:200%;text-align:center">State Based Analysis</h3>


```python
fig = ex.choropleth(locations=e_data.loc[1:,'State'], locationmode="USA-states",color= e_data.loc[1:,'VEP Turnout Rate'], scope="usa",color_continuous_scale=ex.colors.diverging.Portland)

fig.update_layout(title = "VEP Turnout Rate At Each State")
fig.show()
```

<p style="background: transparent; margin-top: 0pt; margin-bottom: 0pt; text-align: center;"><span style='font-family: "Times New Roman", Times, serif; font-size: 24px;'>We can see that Minnesota and Maine have the highest VEP Turnout Rates</span></p>

```python
fig = ex.choropleth(locations=e_data.loc[1:,'State'], locationmode="USA-states",color= e_data.loc[1:,'Voting-Eligible Population (VEP)'], scope="usa",color_continuous_scale=ex.colors.diverging.Portland)

fig.update_layout(title = "Voting-Eligible Population (VEP) At Each State")
fig.show()
```

<p style="background: transparent; margin-top: 0pt; margin-bottom: 0pt; text-align: center;"><span style='font-family: "Times New Roman", Times, serif; font-size: 24px;'>We can see that California is the leading state when it comes to  Voting-Eligible Population</span></p>
