# How is the implementation of existing strategies affecting the rates of COVID-19 infection?


# Introduction
&emsp;&emsp;Coronavirus disease 2019 (COVID-19) is an infectious disease caused by severe acute respiratory syndrome coronavirus 2 (SARS-CoV-2). The disease was first identified in December 2019 in Wuhan, China. Since then, it has spread globally. At first, the World Health Organization (WHO) has declared the outbreak a Public Health Emergency of International Concern (PHEIC) on the 30th of January 2020. Almost two months later, WHO declared it a pandemic. It has been confirmed that local transmission has occurred in most of the world.<br>
&emsp;&emsp;WHO has prescribed recommended measures in an effort to try to contain the spread of the virus. They include frequent hand washing, maintaining social distance from others, covering coughs and sneezes with a tissue or inner elbow, and keeping unwashed hands away from the face.<br>
&emsp;&emsp;At the time of writing, there is no specific antiviral treatment or vaccine available to treat COVID-19. Currently, the treatment focuses on symptom treatment, supportive care, and isolation.<br>
&emsp;&emsp;Ever since the COVID-19 started it’s rapid spread, the world has ground to a halt. To help flatten the curve, governments around the world have started closing businesses, cities, and in some cases even whole regions. Furthermore, they started issuing stay at home orders which exclude only essential workers. All of this led us to an economic slowdown which no-one expected.<br>
&emsp;&emsp;The goal of the project is to find out how the implementations of these strategics affected the rates of COVID-19 infection. The results of this analysis can help governments and organizations around the world by showing what impact measures have on the spread. This will help them stop the spread of the virus by balancing between harsher and lenient measures while at the same time minimizing the impact on the economy.<br>
&emsp;&emsp;Besides, the results of this research could help us be prepared for any similar events in the future. If we were to know the most efficient measures to put in place at each moment in time based on the current situation, we would be able to solve any world crisis of this type without difficulties while maintaining our normal lifestyle as much as we could.

# Data
&emsp;&emsp;Multiple data sources will be used for analysis in this project. The datasets were chosen based on credibility and their target use cases.<br>
&emsp;&emsp;For the overall number of confirmed cases and deaths across the world, the analysis will be performed on The European Centre for Disease Prevention and Control (ECDC) dataset. The ECDC collects the number of COVID-19 cases and deaths, based on reports from health authorities worldwide. This comprehensive and systematic process is carried out daily. It includes information from 206 countries around the globe.<br>
&emsp;&emsp;To help analyze the impact on people’s lives, we are going to use Google’s Community Mobility Reports data. These reports provide insights into what has changed in response to policies aimed at combating COVID-19. They show us movement trends over time across different categories of places such as retail and recreation, groceries and pharmacies, parks, transit stations, workplaces, and residential.<br>
&emsp;&emsp;The insights on various government measures implemented in response to the pandemic are found in the ACAPS COVID-19 dataset. This gives us information for 196 different countries about what type of measure was implemented, when, by whom, and what the prescribed fines are for disobedience. The project will focus on these measures to analyze what impact they had on the rates of infection.

# Analysis
&emsp;&emsp;The analysis was started by making the graphs for total amounts of cases and deaths in respective regions at each point in time. To make exponential growth easier to see, both graphs with linear and logarithmic scales were created.

&emsp;&emsp;Based on the above graphs we can make the following observations:<br>
* First cases and rapid spread come from China as the epicenter of the epidemic where it all started.
* Each region has, at some point, had exponential growth of cases and deaths.
* Europe and the United States have, in March, become regions with the highest number of confirmed cases.
* Most of the regions, except China, are currently experiencing linear growth in COVID-19 cases.
* China is the only region that has managed to put an end to the rapid spread of the virus and is currently encountering only a handful of cases each day.<br>
&emsp;&emsp;Implemented measures are split into 6 categories: public health measures, movement restrictions, social distancing, governance and socio-economic measures, and lockdown. The following graphs will show us the breakdown of the extent of the mentioned measures across the selected regions.

![](https://github.com/PatrikDurdevic/How-is-the-implementation-of-existing-strategies-affecting-the-rates-of-COVID-19-infection/raw/master/ipynb%20data/fig2.png)
![](https://github.com/PatrikDurdevic/How-is-the-implementation-of-existing-strategies-affecting-the-rates-of-COVID-19-infection/raw/master/ipynb%20data/fig3.png)

&emsp;&emsp;Graphs clearly show that the majority of measures fall into the category of public health measures, followed by movement restrictions, social distancing and governance and socio-economic measures. The least amount of measures can be seen in the lockdown category. When looking into what regions implemented which measures, some interesting observations can be made.<br>
&emsp;&emsp;In Africa, governments are putting equal emphasis on almost all of the measures including public health measures, social distancing, movement restrictions, and governance and socio-economic measures. Lockdown being the one measure they almost did not have to use.<br>
&emsp;&emsp;What’s very important to discern is the American (United States) tendency towards public health and governance and socio-economic measures. There are far more of them compared to movement restrictions and social distancing. Perhaps, in the United States, economic situation and personal freedom play a much bigger role than anywhere else.<br>
&emsp;&emsp;Europe is emphasizing on public health measures and social distancing without being afraid of implementing some movement restriction measures. The smaller number of these measures can be attributed to one of the European Union’s fundamental principles which is the free movement of people.<br>
&emsp;&emsp;On the contrary, Middle Eastern countries favor movement restrictions compared to other available measures. They have implemented more movement restriction measures than any other region.<br>
&emsp;&emsp;Unfortunately, this data can not be taken completely for granted as for one completely thought out social distancing measure in one country can make up for numerous leaner measures in another. Having that in mind, the following part of the analysis will give us more robust information about the effectiveness of the measures.<br>
&emsp;&emsp;The most talked-about measure countries are implementing to prevent the spread of COVID-19 is definitely social distancing. By comparing the data from Google mobility reports, and daily percentage increases of cases in various regions, we will analyze how effective it was to limit public gatherings. This part of the analysis will be performed on individual countries since the measures are put in place at different times across the world. Let’s take a look at how placing restrictions on public gatherings impacted pandemic growth.<br>
&emsp;&emsp;The first country analyzed is the United States. The United States has implemented measures that limit public gatherings on two occasions, March 11th, and March 24th respectively. The US was chosen as the first country to be analyzed due to the plethora of accurate Google mobility data. The following graphs show us exactly how people responded when the government put in the above-mentioned measures.


![](https://github.com/PatrikDurdevic/How-is-the-implementation-of-existing-strategies-affecting-the-rates-of-COVID-19-infection/raw/master/ipynb%20data/fig4.png)

&emsp;&emsp;What can be seen is that immediately after the first measure was implemented, the numbers for retail, recreation, transit, and workplaces fell by up to about 40%. On the contrary, people started spending almost 20% more time in their homes. At the same time, grocery and park numbers stayed around the baseline.<br>
&emsp;&emsp;The conclusion is that these measures have indeed made an impact on life in the United States. People started going to work and retail stores less and less, while at the same time they started spending more time at home following the government’s instructions. It is understandable that there was not much of a change for groceries since people still need to go buy essentials for survival. The only unfortunate piece of information that can be found here is that people continued to spend the same amount of time in parks as they have before.<br>
&emsp;&emsp;The following graph will display how the rates of COVID-19 infections changed upon the implementation of said measures.

![](https://github.com/PatrikDurdevic/How-is-the-implementation-of-existing-strategies-affecting-the-rates-of-COVID-19-infection/raw/master/ipynb%20data/fig5.png)

&emsp;&emsp;Immediately after the implementation of measures, not much has changed. That is because we are looking for information in the wrong place. According to Worldometer, the incubation period of COVID-19 ranges from 2 to 14 days. Knowing that, let’s look at the graph once again, but this time looking for the changes following the 14 day period (the red line). It is clear that the percentage change of COVID-19 cases started dropping from 0.5% all the way down to about 0.1%. Exact changes for the can be seen in the table below.

![](https://github.com/PatrikDurdevic/How-is-the-implementation-of-existing-strategies-affecting-the-rates-of-COVID-19-infection/raw/master/ipynb%20data/fig6.png)

&emsp;&emsp;The conclusion for the United States is that measures limiting public gatherings have made a big impact on people’s lives which has, in turn, slashed the growth of COVID-19 cases.<br>
&emsp;&emsp;The next country worth analyzing is Italy, which has been the worst-hit country in Europe. Up until now, they have implemented 5 different measures limiting public gatherings.<br>

![](https://github.com/PatrikDurdevic/How-is-the-implementation-of-existing-strategies-affecting-the-rates-of-COVID-19-infection/raw/master/ipynb%20data/fig7.png)

&emsp;&emsp;It is clear that the first measure had not made nearly an impact which the second one did. Taking a deeper dive into the available data, it is found that the first measure only included the closure of museums and cancellations of educational trips. On the contrary, the second measure outright banned public gatherings such as sports or any other events.

![](https://raw.githubusercontent.com/PatrikDurdevic/How-is-the-implementation-of-existing-strategies-affecting-the-rates-of-COVID-19-infection/master/ipynb%20data/fig8.png)

&emsp;&emsp;Looking at the numbers for the mentioned measures, while the first measure made an impact, the second measure is the one which really changed how people behave nowadays in Italy, as some categories dropped by over 50%. In addition, the graph tells us that these values continued dropping for over 7 days after the second measure while the population started adapting to the new way of life. Let’s take a look at how all of this translated into the number of COVID-19 cases.

![](https://raw.githubusercontent.com/PatrikDurdevic/How-is-the-implementation-of-existing-strategies-affecting-the-rates-of-COVID-19-infection/master/ipynb%20data/fig9.png)

&emsp;&emsp;The graph shows us that even though the second measure made a bigger impact on people’s lives, the first measure is the one that comes on top when comparing cases. Nevertheless, both the first and second measures impacted the growth of cases in a good way by reducing the percentage change of not only cases but deaths too.

![](https://raw.githubusercontent.com/PatrikDurdevic/How-is-the-implementation-of-existing-strategies-affecting-the-rates-of-COVID-19-infection/master/ipynb%20data/fig10.png)

&emsp;&emsp;The table of exact values of changes confirms the observations, both measures made quite an impact. The big change following the first measure can be attributed to the huge growth spurt of well over 100% at the beginning of the pandemic in the country.<br>
&emsp;&emsp;Creating similar graphs and analysis for other countries leads us to the same conclusions. What’s left to analyze is the special situation in Sweden whose lenient measures are making it interesting to analyze. They have implemented 3 measures limiting public gathering in total:
* Limit of public gatherings to 500 persons.
* Restaurants, bars, and cafés throughout the entire country need to take special measures to reduce the risk of spreading COVID-19.
* Max 50 person gatherings in public.<br><br>
Before making any assumptions about how this kind of approach to the pandemic is good or bad, let’s see the graphs.


![](https://raw.githubusercontent.com/PatrikDurdevic/How-is-the-implementation-of-existing-strategies-affecting-the-rates-of-COVID-19-infection/master/ipynb%20data/fig11.png)

&emsp;&emsp;Google mobility reports tell us an expected story, even though changes are present, they are not nearly as drastic as in Italy. Interestingly enough, people have even started spending more and more time in parks. Taking a look at the cases, we can see that even just the first measure alone managed to mitigate most of the growth present in the country. The table will tell us the complete story.

![](https://raw.githubusercontent.com/PatrikDurdevic/How-is-the-implementation-of-existing-strategies-affecting-the-rates-of-COVID-19-infection/master/ipynb%20data/fig12.png)

&emsp;&emsp;Even though the first measure decreased the daily growth percentage by above 30%, both the second and third measures managed to play their part and decrease the growth by about 5%. We can conclude that Sweden has indeed managed to slow the spread of the virus with its lenient measures. What would the numbers look like had they implemented stricter measures is not easy to tell due to the wide range of factors affecting the spread.
