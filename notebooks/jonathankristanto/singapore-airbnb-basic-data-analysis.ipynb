## 2. Programming

### a) Find maximum and minimum value

First, I extract all the column which are numbers (data types = int / float)

```python
number_df = df.select_dtypes(np.number)
number_df.head()
```

Iterate the dataframe to get all min & max value, then insert it into an np.array

```python
list_label = []
list_min_val = []
list_max_val = []
for column in number_df.columns:
    list_label += [column]
    list_min_val += [number_df[column].min()]
    list_max_val += [number_df[column].max()]
    
array_min_max = np.array([list_label, list_min_val, list_max_val])
array_min_max
```

Create the final dataframe

```python
df_min_max = pd.DataFrame(array_min_max[1:], columns=array_min_max[0,:])
df_min_max.index = ['Min Value', 'Max Value']
df_min_max
```

### b) Eliminate null value

First let's see which column has missing value

```python
print("Dataframe shape")
print(df.shape)
print("\nData count")
print(df.count())
```

Here we can see that the dataframe should have 7907 rows of data, and the column **name, last_review, reviews_per_month** are missing some values

```python
df[df['name'].isna()]
```

View whether all missing value from last_review column correspond with reviews_per_month column

```python
df[df['last_review'].isnull()].head()
```

```python
df[df['last_review'].isnull() & ~df['reviews_per_month'].isnull()].head()
```

And now we eliminate the null value

```python
# Drop all rows that contain null value
df_dropna_row = df.dropna()
print("Dataframe shape after dropping all rows with null value", df_dropna_row.shape)

# Drop all columns that contain null value
df_dropna_col = df.dropna(axis=1)
print("Dataframe shape after dropping all rows with null value", df_dropna_col.shape)
```

```python
df_fill = df.copy()
# Fill missing value in name
df_fill['name'].fillna("Bejo", inplace=True)

# Fill missing value in last_review
import datetime
x = datetime.date(2015, 12, 13)
df_fill['last_review'].fillna(x, inplace=True)

# Fill missing value in reviews_per_month
import datetime
x = datetime.datetime.now()
df_fill['reviews_per_month'].fillna(df_fill['reviews_per_month'].mean(), inplace=True)

# View Result
df_fill.loc[[1024,1331], :]
```
