```python
import plotly.figure_factory as ff  
def createTable(dataframe,countriesList,columnName):
    df2 = dataframe[dataframe['Country name'].isin(countriesList)]
    df2 = df2.groupby("Country name").mean().reset_index()
    df_life= df2[['Country name',columnName]]
    df_life = df_life.sort_values(by = [columnName, 'Country name'], ascending = False)
    df_life["Rank"] = [i for i in range(1,len(df_life)+1)]
    table = ff.create_table(df_life)
    table.show()
    
def line(dataframe,countriesList,columnName):
    temp_df = dataframe[dataframe["Country name"].isin(countriesList)] 
    px.line(temp_df,x="year",y=columnName,color="Country name",template="plotly_dark").show()
```
