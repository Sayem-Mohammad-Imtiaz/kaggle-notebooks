```python
from sklearn.preprocessing import LabelEncoder
enc =LabelEncoder()
df['gender']=enc.fit_transform(df['gender'])
df['smoking_status']=enc.fit_transform(df['smoking_status'])
df['work_type']=enc.fit_transform(df['work_type'])
df['Residence_type']=enc.fit_transform(df['Residence_type'])
df['ever_married']=enc.fit_transform(df['ever_married'])
df.head()
```

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
print(X_train.shape,X_test.shape)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=0)
X_train, y_train = sm.fit_resample(X_train, y_train.ravel())
print(X_train.shape,X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from lazypredict.Supervised import LazyClassifier
clf = LazyClassifier(predictions=True)
models, predictions = clf.fit(X_train, X_test, y_train,y_test)
models
```

**From the above we can see the models which give the highest accuracy **

```python
from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier
from xgboost import XGBClassifier
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
xg = XGBClassifier()
xg.fit(X_train, y_train)
ad = AdaBoostClassifier(n_estimators = 50)
ad.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report as cr
from sklearn.metrics import accuracy_score as As
print("XGB Model")
print(cr(y_test,xg.predict(X_test)))
print("Ada Boost Model")
print(cr(y_test,ad.predict(X_test)))
print("Random Forest Model")
print(cr(y_test,rf.predict(X_test)))
```
