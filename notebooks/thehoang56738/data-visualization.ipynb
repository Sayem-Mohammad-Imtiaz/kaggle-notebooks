```python
print(__doc__)

# Author: Kemal Eren <kemal@kemaleren.com>
# License: BSD 3 clause

import numpy as np
from matplotlib import pyplot as plt

from sklearn.datasets import make_biclusters
from sklearn.datasets import samples_generator as sg
from sklearn.cluster.bicluster import SpectralCoclustering
from sklearn.metrics import consensus_score

data, rows, columns = make_biclusters(
    shape=(300, 300), n_clusters=5, noise=5,
    shuffle=False, random_state=0)

plt.matshow(data, cmap=plt.cm.Blues)
plt.title("Original dataset")

data, row_idx, col_idx = sg._shuffle(data, random_state=0)
plt.matshow(data, cmap=plt.cm.Blues)
plt.title("Shuffled dataset")

model = SpectralCoclustering(n_clusters=5, random_state=0)
model.fit(data)
score = consensus_score(model.biclusters_,
                        (rows[:, row_idx], columns[:, col_idx]))

print("consensus score: {:.3f}".format(score))

fit_data = data[np.argsort(model.row_labels_)]
fit_data = fit_data[:, np.argsort(model.column_labels_)]
plt.matshow(fit_data, cmap=plt.cm.Blues)
plt.title("After biclustering; rearranged to show biclusters")

plt.show()
```

```python
for sub in range(1,len(total_teacher_per_sub)+1):
    #filter teachers_id each subject
    teachers_id = iss_teacher_sub[iss_teacher_sub['subject_id']==sub]
    teachers_id = teachers_id['teacher_id'].tolist()
    
    #filter degree of teacher each subject
    degree_teachers= iss_teacher[iss_teacher['id'].isin(teachers_id)]
    degree_teachers = degree_teachers['degree']
    temp = pd.Series(degree_teachers.isnull().sum(), index=['null'])
    total_degree = degree_teachers.value_counts().append(temp)
    
    degree = iss_teacher['degree'].value_counts().sort_index()
    degree[[i for i in range(len(degree))]] = 0
    degree = degree.append(pd.Series(0,index=['null']))

    degree[[total_degree.index[i] for i in range(len(total_degree))]] = total_degree.tolist()
    for i in range(len(degree)):
        raw_data[degree.index[i]].append(degree[i])
raw_data
```

```python
# Data
r = [i for i in range(len(total_teacher_per_sub))]
df = pd.DataFrame(raw_data)
 
# From raw value to percentage
totals = [i+j+k+l for i,j,k,l in zip(df['Cá»­ nhÃ¢n'], df['Tháº¡c sÄ©'], df['Tiáº¿n sÄ©'],df['null'])]
greenBars = [i / j * 100 for i,j in zip(df['Cá»­ nhÃ¢n'], totals)]
print(greenBars)
orangeBars = [i / j * 100 for i,j in zip(df['Tháº¡c sÄ©'], totals)]
print(orangeBars)
if df['Tiáº¿n sÄ©'].sum() == 0:
    blueBars = [0 for i in range(len(total_teacher_per_sub))]
else:
    blueBars = [i / j * 100 for i,j in zip(df['Tiáº¿n sÄ©'], totals)]
print(blueBars)
blackBars = [i / j * 100 for i,j in zip(df['null'], totals)]
print(blackBars)

# plot
barWidth = 0.85
names = subjects[0:8]
plt.figure(figsize=(12,6))
# Create green Bars
plt.bar(r, greenBars, color='r', edgecolor='white', width=barWidth)
# Create orange Bars
plt.bar(r, orangeBars, bottom=greenBars, color='b', edgecolor='white', width=barWidth)
# Create blue Bars
plt.bar(r, blueBars, bottom=[i+j for i,j in zip(greenBars, orangeBars)], color='#000000',
        edgecolor='white', width=barWidth)
plt.bar(r, blackBars, bottom=[i+j+k for i,j,k in zip(greenBars, orangeBars,blueBars)], color='g',
        edgecolor='white', width=barWidth)
 
# Custom x axis
plt.xticks(r, names)
plt.xlabel("TrÃ¬nh Äá» cá»§a cÃ¡c giÃ¡o viÃªn trong tá»«ng mÃ´n há»c (dá»¯ liá»u iss)")
plt.legend(labels=df.columns) 
# Show graphic
plt.show()
```
