# Выводы

Данный датасет посвящен потреблению алкоголя, поэтому смотрим на показатели Dalc (ежедневное употребление) и Walc (недельное употребление).

Наблюдается корреляция употребления алкоголя с количеством пропусков (absences) и проваленных курсов (failures). А также отрицательная корреляция употребления алкоголя со количеством времени, посвящаемым учебе (studytime). Возможно, все вещи имеют общую причину. Возможно, одна из них является причиной других.
