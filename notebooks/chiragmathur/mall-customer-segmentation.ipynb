```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
```

```python
data = pd.read_csv("../input/customer-segmentation-tutorial-in-python/Mall_Customers.csv")
```

```python
data.rename(columns={'Annual Income (k$)':'Income','Spending Score (1-100)':'Score'}, inplace = True)
```

```python
data
```

****SCORE VS INCOME****

```python
sns.scatterplot(data = data,x='Score', y='Income')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 5 , init='k-means++')
kmeans = kmeans.fit(data[['Score','Income']])
```

```python
kmeans.cluster_centers_
```

```python
data['cluster'] = kmeans.labels_
```

```python
data
```

```python
sns.scatterplot(data = data,x='Score', y='Income', hue='cluster')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='red',marker='s')
```

****AGE VS SCORE****

```python
sns.scatterplot(data=data,x='Age',y='Score',color='g' )
```

NO. OF CLUSTERS - ELBOW METHOD

```python
k_rng = (1,11)
sse = []
for k in k_rng:
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans = kmeans.fit(data[['Age','Score']])
    sse.append(kmeans.inertia_)
```

```python
kmeans.cluster_centers_
```

```python
sse
```

```python
sns.scatterplot(data = data,x='Age', y='Income', hue='cluster')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],color='red',marker='s')
```
