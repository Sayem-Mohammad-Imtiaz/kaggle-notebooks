```python
#Importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
#Loading dataset
wine = pd.read_csv('../input/winequality-red.csv')
```

```python
wine.shape
```

```python
train,test = train_test_split(wine, test_size=0.2, random_state=42)
train_input_only=train.drop('quality', axis=1)
```

## Corelation Between Input Variables

```python
plt.figure(figsize=(12,12))
corr =train_input_only.corr()
sns.heatmap(corr,cmap=sns.diverging_palette(20, 220, n=200));
```

```python
plt.figure(figsize=(8,8))
plt.scatter(train['citric acid'],train['volatile acidity'],alpha=0.6);
plt.title('Citric acid vs. Volatile Acidity')
plt.xlabel('citric acid');
plt.ylabel('volatile acidity');
```

```python
plt.figure(figsize=(8,8))
plt.scatter(train['fixed acidity'],train['pH'],alpha=0.4);
plt.title('Fixed Acidity vs. pH')
plt.xlabel('fixed acidity');
plt.ylabel('pH');
```

```python
plt.figure(figsize=(8,8))
plt.scatter(train['citric acid'],train['pH'],alpha=0.4);
plt.title('Citric Acid vs. pH')
plt.xlabel('citric acid');
plt.ylabel('pH');
```

```python
plt.figure(figsize=(8,8));
plt.scatter(train['free sulfur dioxide'],train['total sulfur dioxide'],alpha=0.7);
plt.xlabel('free sulfur dioxide');
plt.ylabel('Total Sulfur Dioxide');
plt.title('Free Sulfur Dioxide vs Total Sulfur Dioxide');
```
