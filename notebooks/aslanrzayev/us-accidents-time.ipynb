# At what time do accidents usually occur in the US

# Task Details
Figure out the time that accidents usually occur in the US. This can be done by using the "start time" and "end time" columns.

# Expected Submission
A notebook that outputs the time that accidents usually occur.

# Evaluation
A proper ranking of the times that accidents occur. The time that accidents usually occur should be at the top of the list together with the number of times accidents occurred during that specific time. Other times at lower ranks should follow. This is done to provide a more detailed solution.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
df = pd.read_csv('../input/us-accidents/US_Accidents_Dec20.csv')
df.head()
```

```python
df.isnull().sum()
```

**We are looking for start / end time. As there is no missing values in these columns, we  can continue to work with dataset**

Let's check in which format we have our values

```python
df.info()
```

**We need Start / End time in datetime format, so we need to convert it**

```python
df_time = df.loc[:,['ID', 'Start_Time', 'End_Time']]
df_time['Start_Time'] = pd.to_datetime(df_time['Start_Time'])
df_time['End_Time'] = pd.to_datetime(df_time['End_Time'])
df_time.info()
```

```python
df_time['Start_hour'] = df_time['Start_Time'].dt.hour
hours = df_time.groupby(['Start_hour']).count()
hours
```

```python
f, ax = plt.subplots(figsize = (18, 10))
sns.barplot(x = hours.index, y = 'Start_Time', data = hours)
plt.xlabel('Time of accident (hour)', labelpad = 10, fontsize=12, weight='bold')
plt.ylabel('Number of accidents', labelpad = 10, fontsize=12, weight='bold')
plt.title('Time when accident usually oocur in the US', fontsize = 15, weight = 'bold')
```
