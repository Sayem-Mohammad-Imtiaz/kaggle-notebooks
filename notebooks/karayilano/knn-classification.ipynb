```python
#Making abnormal ones 1 and normal ones 0
df["class"] = [1 if i == "Abnormal" else 0 for i in df["class"]]
```

```python
x = df.drop(["class"], axis=1)
y = df["class"]
```

```python
# normalization process
x = (x - x.min()) / (x.max() - x.min())
```

```python
#train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state = 42)
```

```python
# knn for neighbor value 3
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
knn.score(x_test, y_test)
```

```python
#lets find what is best neighbor value

import matplotlib.pyplot as plt
score_list = []
for i in range(1,100):
    knn2 = KNeighborsClassifier(n_neighbors=i)
    knn2.fit(x_train, y_train)
    score_list.append(knn2.score(x_test, y_test))
    
plt.plot(range(1,100), score_list)
plt.xlabel("neighbor value")
plt.ylabel("score")
plt.show()
```

It is arount 40. Lets find exact number.

```python
import numpy as np
np.argmax(score_list)
```

So best neighbor value is 38 since when you look carefuly into for loop we start with 1 then add them to score_list. 

```python
#lets check
knn = KNeighborsClassifier(n_neighbors=38)
knn.fit(x_train, y_train)
knn.score(x_test, y_test)
```
