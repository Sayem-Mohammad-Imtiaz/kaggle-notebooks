```python
#  Flavor profile for veg and non veg dishes 
def pie_chart_1(df):
    
    veg = df[df['diet']=='vegetarian']['flavor_profile'].value_counts()
    n_veg = df[df['diet']=='non vegetarian']['flavor_profile'].value_counts()

    
    
    fig = make_subplots(rows=1, cols=2, specs=[[{'type':'domain'}, {'type':'domain'}]])    
    fig.add_trace(go.Pie(labels= veg.index, values= veg.values, name="Veg tastes",title='VEG'),
                  1, 1)
    fig.add_trace(go.Pie(labels= n_veg.index, values = n_veg.values, name="non_veg tastes",title='Non_Veg'),
              1, 2)

    fig.update_traces(hole=.4, hoverinfo="label+percent+name")

    fig.show()
pie_chart_1(df)
```

<a href="#top" class="btn btn-primary btn-sm"  aria-pressed="true" style="color:white" data-toggle="popover" title="go to Colors">Go to menu</a>

```python
# basemap of total dishes distribution 
def basemap_1(df):
    new_df = df.copy()
    indexes = new_df[new_df['state']=='-1'].index
    new_df.drop(indexes,axis=0,inplace=True)
    x = new_df['state'].value_counts()
    state_names = x.index
    state_count = x.values
    
    
    lat,lon = [],[]
    for i in range(len(state_names)):
        query = state_names[i]
        results = geocoder.geocode(query)
        lat.append(results[0]['geometry']['lat'])
        lon.append(results[0]['geometry']['lng'])
        
    
    col1 = pd.DataFrame(state_names,columns=['state_name'])
    col2 = pd.DataFrame(lat,columns=['lat'])
    col3 = pd.DataFrame(lon,columns=['lon'])
    col4 = pd.DataFrame(state_count,columns=['count'])
    
    india_df = pd.concat([col1,col2,col3,col4],axis=1)
    
    

    fig = px.scatter_mapbox(india_df, lat="lat", lon="lon", size='count',hover_name="state_name",
                            color_discrete_sequence=["fuchsia"], zoom=3.5, height=500,width=500)
    fig.update_layout(mapbox_style="open-street-map")
    fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
    fig.show()
basemap_1(df)   
```
