판다스에서는 데이터를 `Series` 또는 `DataFrame` 으로 취급
- `Series`: 1차원 배열
- `DataFrame`: 2차원 배열

기본적으로는 `DataFrame`이 많이 사용됨.

`pd.read_csv` 함수: 말 그대로 CSV 파일 읽어옴. 읽힌 데이터는 `DataFrame`으로 래핑
`DataFrame.describe` 메서드: 데이터 전반의 통계적 수치를 한눈에 파악

```python
melbourne_file_path = '../input/melbourne-housing-snapshot/melb_data.csv'

# CSV 파일 읽고 => DataFrame으로 변환
melbourne_data = pd.read_csv(melbourne_file_path) 

# 요약 정보 출력
melbourne_data.describe()
```
