```python
fy_date_dict = {}

for fy in fy_area_dict:
    
    fy_date_dict[fy]={}

    for area in fy_area_dict[fy]:

        info = fy_area_dict[fy][area]

        dates = [month[0] for month in info]
        vals = [month[1:] for month in info]
        

        fy_date_dict[fy][area] = dict(zip(dates,vals)) 
```

```python
def create_table(title, contents, conn,drop=True):
    
    if drop:
        with conn:
            sql = 'DROP TABLE {} ;'.format(title)
            curs=conn.cursor()
            curs.execute(sql)
        
    
    with conn:
        sql = 'CREATE TABLE {} ({});'.format(title,contents)
        curs=conn.cursor()
        curs.execute(sql)
```

```python
title = 'Regions'
contents = ''' Region_ID INT PRIMARY KEY NOT NULL,
                Region TEXT'''
```

```python
create_table(title,contents,conn,drop=delete_dbs)
```

```python
def insert_vals(title,vals,conn):
    

    q_num = ','.join('?'*len(vals[0]))
    sql = 'INSERT INTO {} VALUES({});'.format(title,q_num)
    
    with conn:
        curs = conn.cursor()
        curs.executemany(sql,vals)
```
