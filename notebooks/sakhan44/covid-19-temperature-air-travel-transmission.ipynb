```python
raw_data = pd.read_csv('../input/corona-virus-report/covid_19_clean_complete.csv',parse_dates=True)
country_master=raw_data['Country/Region'].unique().tolist()
```

```python
 data = pd.read_csv('../input/corona-virus-report/covid_19_clean_complete.csv',parse_dates=True)
# data.drop(['WHO Region'])
data = data[['Country/Region','Date','Confirmed','Deaths','Recovered']]
# data[data['Country/Region']=='India']
data['DateNTime'] = pd.to_datetime(data['Date'])
data['dayOfWeek'] = data['DateNTime'].dt.day_name()
dr=data[(data['dayOfWeek']=='Friday')]
# data.groupby("Country/Region")
dr
```

```python
consolidated_data=pd.pivot_table(dr,index=["DateNTime"],values=["Confirmed"],columns=["Country/Region"])
```

```python
consolidated_data.columns=country_master
consolidated_data['ID'] = range(1, 1+len(consolidated_data))
```

```python
plt.figure(figsize=(22.5,11.2))

y = consolidated_data.index.sort_values()

# India
plt.subplot(221)
plt.plot(y,consolidated_data['India'])
#plt.yscale('log')
plt.title('India')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
# Italy
plt.subplot(222)
plt.plot(y,consolidated_data['Italy'])
plt.yscale('log')
plt.title('Spain')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
# Spain
plt.subplot(223)
plt.plot(y,consolidated_data['Spain'])
plt.yscale('log')
plt.title('Italy')
plt.grid(True)
plt.xticks(rotation=45, ha='right')
# US
plt.subplot(224)
plt.plot( y,consolidated_data['US'])
plt.yscale('log')
plt.title('US')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
X = np.array(consolidated_data['India']).reshape(-1, 1) 
y = np.array(consolidated_data['ID']).reshape(-1, 1) 

# Separating the data into independent and dependent variables 
# Converting each dataframe into a numpy array 
# since each dataframe contains only one column 
#df_binary.dropna(inplace = True) 

# Dropping any rows with Nan values 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25) 

# Splitting the data into training and testing data 
regr = LinearRegression() 

regr.fit(X_train, y_train) 
print(regr.score(X_test, y_test)) 
```

```python
y_pred = regr.predict(X_test) 
plt.scatter(y_test,X_test, color ='b') 
plt.plot(y_pred,X_test, color ='k') 
  
plt.show()
```
