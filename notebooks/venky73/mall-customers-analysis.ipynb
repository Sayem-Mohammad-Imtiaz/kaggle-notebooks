I would like to categorise Age with 5yrs as frequency

```python
def group_age(age):
    if age <= 30:
        return "<= 30"
    elif age <= 40:
        return "31-40"
    elif age<= 50:
        return "41-50"
    elif age<= 60:
        return "51-60"
    else:
        return "61-70"
```

```python
df["Age"]= df.Age.apply(group_age)
```
