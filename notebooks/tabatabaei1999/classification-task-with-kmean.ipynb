```python
l = [3,5,6,7,9,-5]
'''
I select these columns because have two variable and also all ring-number column to this list because 
it is number(read data describtion)
'''
x = df.iloc[:,:].values
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in l:
    x[:,i] = le.fit_transform(x[:,i])
```

```python
ll=[]
for i in range(21):
    if i not in l:
        ll.append(i)

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), ll)], remainder='passthrough', sparse_threshold=0)
x = np.array(ct.fit_transform(x))

y = le.fit_transform(y)
```

I use MinMaxScaler as we have 2 in ring-type after decoding

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

# KMean

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 2, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(x)
```

rand_score

```python
from sklearn.metrics.cluster import rand_score

rand_score(y,kmeans.labels_)
```
