```python
test.drop(['id'],axis =1,inplace=True)
test.shape, train.shape
```

```python
X = train.drop(['price_range'], axis = 1)
y = train['price_range']
```

```python
train.columns
```

```python
cols = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi',]
num_cols =  list(X.drop(cols, axis = 1).columns)


model = StandardScaler()
X[num_cols] = model.fit_transform(X[num_cols])
test[num_cols] = model.transform(test[num_cols])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
models = [LogisticRegression(),RandomForestClassifier(),
          DecisionTreeClassifier(), KNeighborsClassifier(),
         GaussianNB(),SVC(),XGBRegressor(),GradientBoostingClassifier()]
```

```python
def calculate(model):
    
    m =model
    m.fit(X_train,y_train)
    predict =mod.predict(X_test)
    scor=m.score(X_train,y_train)
    #print('accuracy on training set = ',scor)
    return scor
    
    
    #print(classification_report(y_test, predict))

```

```python
calculate(LogisticRegression())
```

```python
scores = []
model_names = []
for mod in models:
    m =mod
    m.fit(X_train,y_train)
    predict =mod.predict(X_test)
    scor=m.score(X_train,y_train)
    scores.append(scor)
    model_names.append(mod.__class__.__name__)
    
    
```

```python

plt.figure(figsize=(12,10))
sns.barplot(model_names, scores)
plt.xticks(rotation=90)

plt.show()
```

```python
df =pd.DataFrame(scores,model_names)
df
```
