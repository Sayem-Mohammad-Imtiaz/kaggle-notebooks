```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
df = pd.read_csv('../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
```

```python
df.head()
```

```python
df.columns = ['name','author','rating','reviews','price','year','genre']
df.info()
```

Let's see price and user rating distribution.

```python
fig = px.box(df, y="rating")
fig.show()
```

The rating has a fairly small spread. 

Let's look at the tail.

```python
df.query('rating < 4')
```

All books are fictions

```python
fig = px.box(df, y="price")
fig.show()
```

Price has more outliers

```python
df.query('price > 50')
```

Let's look at genre distribution

```python
df_genres = df["genre"].value_counts().reset_index()
df_genres.columns = ['genre', 'count']
fig = px.pie(df_genres, values='count', names='genre', title='genres')
fig.show()
```

Non fiction is more popular. Doea the picture change through the years.

```python
df_genres_by_years = pd.pivot_table(df, values='name', index=['year','genre'],  aggfunc ='count', fill_value=0).reset_index()
df_genres_by_years.columns = ['year', 'genre', 'count']
fig = px.bar(df_genres_by_years, x = 'genre', y = 'count', animation_frame="year", color="genre")

fig.show()
```

Only in 2014 fiction leads. In other years non fiction is more or less ahead.

And what about prices?

```python
df_price_by_years = pd.pivot_table(df, values='price', index=['year','genre'],  aggfunc =np.mean, fill_value=0).reset_index()
fig = px.bar(df_price_by_years, x = 'genre', y = 'price', animation_frame="year", color="genre")

fig.show()
```

Mean price for non fiction is higher than for fiction. The most dramatic difference is in 2013-2014. This corresponds to the year of "Diagnostic and Statistical Manual ..."

Let's look at the rating

```python
df_rating_by_years = pd.pivot_table(df, values='rating', index=['year','genre'],  aggfunc =np.mean, fill_value=0).reset_index()
fig = px.bar(df_rating_by_years, x = 'genre', y = 'rating', animation_frame="year", color="genre")

fig.show()
```

We can see that rating is almost undependent of the year and genre.

Do users write more and more reviews?

```python
df_review_by_years = pd.pivot_table(df, values='reviews', index=['year','genre'],  aggfunc =np.sum, fill_value=0).reset_index()
fig = px.bar(df_review_by_years , x = 'genre', y = 'reviews', animation_frame="year", color="genre")

fig.show()
```

Number of reviews grows steadily as people use service more intensively. We can see that till last two years fiction books have more reviews than non fiction ones, but the picture has changed.
