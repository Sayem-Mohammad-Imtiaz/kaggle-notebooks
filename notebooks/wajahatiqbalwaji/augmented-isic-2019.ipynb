```python
data_set.drop('UNK',axis=1,inplace=True)
cols=['NV','BCC','AK','BKL','DF','VASC','SCC','MEL']
```

```python
data_set[data_set[cols[0]]>0].iloc[:3000]

df=pd.concat([data_set[['image','MEL','BCC','AK','BKL','DF','VASC','SCC']], data_set[data_set[cols[0]]>0][cols[0]].iloc[:3000]], axis=1)
df=df.replace(np.nan, 0)
```

```python
train=df


for i in range(0,len(cols)):
      print("Total images of " + cols[i]+ " : "+ str(len(train[train[cols[i]]>0])))
```

```python
w=[2,2,7,2,19,18,8,1]


print(w)


for i in range(0,len(cols)):
      print("Total images of " + cols[i]+ " : "+ str(len(train[train[cols[i]]>0])*w[i]*2))
```

```python
import matplotlib.pyplot as plt
from skimage.transform import resize
datagen = ImageDataGenerator(
        zoom_range=0.3,
      
        width_shift_range=0.1,
        height_shift_range=0.1,
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='nearest')

for z in range(0,len(cols)):
    new_set=train[train[cols[z]]>0]

    for i in range(len(new_set)):


        img = load_img(data_set_path + new_set['image'].iloc[i]+'.jpg')  
        x = img_to_array(img)
        x=resize(x, (1000, 1000))
        save_img(new_data_set_path+"/"+new_set['image'].iloc[i]+".jpg",x)
        x = x.reshape((1,) + x.shape)
       
        
        j = 1
        for batch in datagen.flow(x, batch_size=1):
            f=w[z]
            if z==0 or z==1 or z==7:
                f=w[z]
            else :
                f=w[z]*2
                
            image = batch[0].astype('uint8')
            im_name=new_set['image'].iloc[i]+"_"+str(j)
            save_img(new_data_set_path+"/"+im_name+".jpg",batch[0])
            s=train[train[cols[z]]>0].iloc[i]
            s['image']=im_name


            train=train.append(s,ignore_index = True)
    
            
                
            j += 1
            if j > int(f):
                break  
    print(z)
```
