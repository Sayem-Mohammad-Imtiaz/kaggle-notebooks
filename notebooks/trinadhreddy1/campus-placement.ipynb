```python
from sklearn.model_selection import train_test_split
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
from pandas.plotting import scatter_matrix
```

```python
my_data.loc[my_data['status']=='Not Placed', 'status'] = 0
my_data.loc[my_data['status']=='Placed', 'status'] = 1
my_data.loc[my_data['degree_t']=='Sci&Tech','degree_t'] = 0
my_data.loc[my_data['degree_t']=='Comm&Mgmt','degree_t'] = 1
my_data.loc[my_data['degree_t']=='Others','degree_t'] = 2
my_data.loc[my_data['gender']=='M','gender'] = 0
my_data.loc[my_data['gender']=='F','gender'] = 1
my_data.loc[my_data['ssc_b']=='Others','ssc_b'] = 0
my_data.loc[my_data['ssc_b']=='Central','ssc_b'] = 1
my_data.loc[my_data['hsc_b']=='Others','hsc_b'] = 0
my_data.loc[my_data['hsc_b']=='Central','hsc_b'] = 1
my_data.loc[my_data['ssc_b']=='Others','ssc_b'] = 0
my_data.loc[my_data['ssc_b']=='Central','ssc_b'] = 1
my_data.loc[my_data['workex']=='No','workex'] = 0
my_data.loc[my_data['workex']=='Yes','workex'] = 1
my_data.loc[my_data['hsc_s']=='Commerce','hsc_s']=0
my_data.loc[my_data['hsc_s']=='Science','hsc_s']=1
my_data.loc[my_data['hsc_s']=='Arts','hsc_s']=2
my_data.loc[my_data['specialisation']=='Mkt&HR','specialisation']=0
my_data.loc[my_data['specialisation']=='Mkt&Fin','specialisation']=1


my_data['workex'].astype(int)
my_data['ssc_b'].astype(int)
my_data['hsc_b'].astype(int)
my_data['gender'].astype(int)
my_data['degree_t'].astype(int)
my_data['hsc_s'].astype(int)
my_data['specialisation'].astype(int)

```

```python
x = my_data.drop(['status','sl_no'],axis=1)
y=my_data['status'].astype(int)
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=3)
```

```python
x=x.fillna(value=0,axis=1)
```

```python
x.head()
```

```python
x.isnull().any().sum()
```

```python
y
```

```python
seed=7
models = []
models.append(('RF',RandomForestClassifier()))
models.append(('SVM',SVC()))
models.append(('LR',LogisticRegression()))
models.append(('NB',GaussianNB()))
# Evaluating each models in turn
results = []
names = []
for name, model in models:
    kfold = model_selection.KFold(n_splits=13,random_state=seed)
    cv_results = cross_val_score(model, x_train, y_train, cv=kfold, scoring='accuracy')
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" %(name, cv_results.mean(), cv_results.std())
    print(msg)
```

**Logistic Regression**

```python
logistic = LogisticRegression()
logistic.fit(x_train,y_train)
y_pred=logistic.predict(x_test)
print(classification_report(y_test,y_pred))
accuracy=logistic.score(x_test,y_test)
print (accuracy*100,'%')
```

**SVM**

```python
classifier2=SVC()
classifier2.fit(x_train,y_train)
y_pred=classifier2.predict(x_test)
print(classification_report(y_test,y_pred))
accuracy=classifier2.score(x_test,y_test)
print(accuracy*100 , '%')
```

**Randon forest classifier**

```python
classifier3=RandomForestClassifier()
classifier3.fit(x_train,y_train)
y_predict=classifier3.predict(x_test)
print(classification_report(y_test,y_predict))
accuracy=classifier3.score(x_test,y_test)
print(accuracy*100,'%')
```

GaussianNB

```python
classifier4=GaussianNB()
classifier4.fit(x_train,y_train)
y_pred=classifier4.predict(x_test)
print(classification_report(y_test,y_pred))
accuracy=classifier4.score(x_test,y_test)
print(accuracy*100,'%')
```

DecisionTreeClassifier

```python
classifier5=DecisionTreeClassifier()
classifier5.fit(x_train,y_train)
y_pred=classifier5.predict(x_test)
print(classification_report(y_test,y_pred))
accuracy=classifier5.score(x_test,y_test)
print(accuracy*100,'%')
```
