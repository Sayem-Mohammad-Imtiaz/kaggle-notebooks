```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(10, 8)}); 
```

Загрузите данные. Они находятся в директории input.

```python
df = pd.read_csv('../input/zomato.csv', encoding='latin1')
```

В каких странах расположены рестораны? Назовите топ-5 стран. Выведите аналогичную статистику по городам.

Какие типы кухни подают в ресторанах? Как связан этот признак с рейтингом? Не забывайте о визуализации информации.

```python
df.groupby('Cuisines')['Aggregate rating'].mean()
```

Что можно сказать о связи ценового диапазона и рейтинга ресторана? Не забывайте о визуализации информации.

```python
df.groupby('Aggregate rating')['Average Cost for two'].mean().plot(kind='bar')
plt.ylabel('Average Cost for two') 
plt.show();
```

Есть ли различия в рейтинге ресторанов с возможностью бронирования столиков и без него? 

```python
df.groupby('Has Table booking')['Aggregate rating'].mean().plot(kind='bar')
plt.ylabel('Aggregate rating') 
plt.show();
```

У ресторанов с возможностью заказать столик рейтинг в среднем выше.

Визуализируйте признак **Rating color** (или **Rating text**) в комбинации с геолокацией (**Longitude**, **Latitude**). Какой вывод можно сделать?

```python
df.groupby('Rating text')['Longitude', 'Latitude'].mean().plot(kind='bar')
plt.ylabel('Longitude and Latitude') 
plt.show();
```

Чем ниже геолокация, тем выше рейтинг.

Какие ещё инсайты можно найти в имеющихся данных? Творческое задание.

Как зависит рейтинг ресторана от возможности заказать онлайн?

```python
df.groupby('Has Online delivery')['Aggregate rating'].mean().plot(kind='bar')
plt.ylabel('Aggregate rating') 
plt.show();
```

Очевидный результат: у ретсоранов с доставкой рейтинг выше почти на балл.

На сколько часто люди оставляют хорошие и плохие отзывы?

```python
df.groupby('Aggregate rating')['Votes'].mean().plot(kind='bar')
plt.ylabel('Votes') 
plt.show();
```

Как можно видеть, люди не особо любят "ругать" рестораны.
