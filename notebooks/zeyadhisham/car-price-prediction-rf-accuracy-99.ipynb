# Cleaning Data

* **Dropping unused features**

```python
df = df.drop(['Unnamed: 0', 'vin', 'lot', 'condition'], axis=1)
df.head()
```

* **Encoding Data**

```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
df["brand_num"] = lb_make.fit_transform(df["brand"])
df["model_num"] = lb_make.fit_transform(df["model"])
df["title_num"] = lb_make.fit_transform(df["title_status"])
df["color_num"] = lb_make.fit_transform(df["color"])
df["state_num"] = lb_make.fit_transform(df["state"])
df["country_num"] = lb_make.fit_transform(df["country"])
df.head()
```

# Train Model

> Decision Tree

```python
from sklearn import tree

dt = tree.DecisionTreeClassifier()

feature_names = ['brand_num',	'model_num',	'title_num',	'color_num'	,'state_num'	,'country_num','mileage', 'year']
target_name = ['price']

dt.fit(
    df[feature_names],
    df[target_name]
)
```

> Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=40, max_depth=20)
rf.fit(
    df[feature_names],
    df[target_name]
)
```

```python
rf.predict([df.iloc[2000][feature_names]])
```

```python
df.iloc[2000]
```

```python
import random
x = random.randint(0,2499)
real_value = df.iloc[x]['price']
predicted_value = rf.predict([df.iloc[x][feature_names]])
acc = (predicted_value/real_value)*100
print(f'The real price of choosen car was {real_value}\nThe predicted value of choosen car was {predicted_value}\nAnd the accuracy was {acc}')
```

```python
predictions = rf.predict(df[feature_names])
```

```python
predictions
```

# Evaluate The Model

```python
from sklearn import metrics
import numpy as np
print('Mean Absolute Error:', metrics.mean_absolute_error(df['price'], predictions))
print('Mean Squared Error:', metrics.mean_squared_error(df['price'], predictions))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(df['price'], predictions)))
```

```python
from sklearn.metrics import r2_score
r2_score(df['price'],predictions) 
```
