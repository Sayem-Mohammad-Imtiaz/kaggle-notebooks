**Let's see the correlation between the data.**

* If value -1, there is a full negative linear relationship. (inverse ratio)
* If value +1, there is a full positive linear relationship. (direct ratio)
* If value  0, there is no relationship between the two variables.
----
* 0.00 - no relationship
* 0.01 - 0.29 low level relationship
* 0.30 - 0.70 medium level relationship
* 0.71 - 0.99 high level of relationship
* 1.00 - Perfect relationship

```python
data.corr()
```

For example, the ratio between attack and defense is 0.438687. Medium level.

Show the relationship between attack and defender with scatter graph.

```python
sns.lmplot(x="Attack",y="Defense", fit_reg=False, hue='Generation', data=data)
# x="Attack",y="Defense" : We write x and y axis tags. 
# fit_reg=False : We remove the regression line.
# hue='Generation': We color according to the generation.
# data=data : We determine the dataset.
plt.ylim(0, None) # We start 0 of y-axis value.
plt.xlim(0, None) # We start 0 of x-axis value.
plt.show() 
```

Show the battle features of Pokemon on the box graph.

```python
feature_data = data.drop(['Total', 'Generation', 'Legendary'], axis = 1) # We have removed 3 columns. Because it's not war values.
sns.boxplot(data = feature_data, notch = True, linewidth = 0.5, width = 0.6)
# notch: With this argument we show the limits of the media in the 95% confidence interval.
# linewidth: With this argument, we determine the line thickness of each box.
# width: With this argument we determine the width of each box.
plt.show()
```

Violin Graph
Show the distribution of Pokemon attack values according to Type 1.

```python
f,ax=plt.subplots(figsize=(15,15)) #boyut ayarlÄ±yoruz
pkmn_type_colors = ['#78C850',  # Grass
                    '#F08030',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#C03028',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                    '#390072',  # Dark
                    '#6a776c',  # Steel
                    '#5dc6c3',  # Flying
                   ] # We have set a special color according to the type of Pokemon.
sns.set_style("whitegrid")
sns.violinplot(x="Type 1",y="Attack",data=data,fmt='.1f',ax=ax, linewidth = 1.0, palette=pkmn_type_colors)

plt.show()
```

By looking at this diagram, it can be said that the attack values are mainly between 50 and 100.

Swarm Graph
Show the distribution of Pokemon attack values according to Type 1.

```python
sns.set_style("whitegrid")
#sns.swarmplot(x="Type 1",y="Attack",data=data, palette=pkmn_type_colors, size=3)
sns.catplot(x="Type 1",y="Attack",data=data, kind="swarm", palette=pkmn_type_colors, height=6, aspect=4)
plt.show()
```

**Merge charts**

```python
sns.violinplot(x = 'Type 1', y = 'Attack', data = data, inner = None, palette=pkmn_type_colors)
sns.swarmplot(x = 'Type 1', y = 'Attack', data = data, color = 'k', alpha = 0.7)
plt.title('Attack by Type')
plt.show()
```

In the chart above, we can see the distribution of Attack values of all Pokemon by type.

**Melt Function**

```python
melted_df = pd.melt(data, 
                    id_vars=["Name", "Type 1", "Type 2"], # Variables to keep
                    var_name="Stat") # Name of melted variable
melted_df.head()
```

As you can see, we have melted all the variables except "Name", "Type 1", "Type 2" into a single column.

Let's print the dimensions of the DataFrames:

```python
print( data.shape )
print( melted_df.shape )
```

The melted_df table is 9 times the data table.

We can make a swarm chart.

```python
plt.figure(figsize=(10,6))
 
sns.swarmplot(x='Stat', y='value', data=melted_df, hue='Type 1', dodge=True, palette=pkmn_type_colors)
 
plt.ylim(0, 260)
 
plt.legend(bbox_to_anchor=(1, 1), loc=2)
plt.show()
```

**Heat Map**

Heat maps are used as a matrix to visualize data.

```python
corr = data.corr()

sns.heatmap(corr)

plt.show()
```

**Histogram**

The histogram is used to visualize the distribution of numerical variables.

```python
sns.distplot(data.Attack)
plt.show()
```

**Column Chart**

The column graph is used to visualize categorical variables.

```python
sns.countplot(x='Type 1', data=data, palette=pkmn_type_colors)
 
# X eksenindeki etiketleri dÃ¶ndÃ¼r
plt.xticks(rotation=-45)
plt.show()
```

**Factor Chart**

The factor graph allows you to easily separate graphs of factor classes.

```python
g = sns.factorplot(x = 'Type 1', y = 'Attack', data = data, hue = 'Generation', col = 'Generation', kind = 'swarm')
g.set_xticklabels(rotation = -45)
plt.show()
```

**Density Chart**

The density graph shows the distribution between the two variables.

```python
sns.kdeplot(data.Attack, data.Defense)
plt.show()
```

**Common Distribution Graph**

The common scatter plot combines the scatter plot and histogram to give us detailed information about the two-variable distributions.

```python
sns.jointplot(x = 'Attack', y = 'Defense', data = data)
plt.show()
```
