Importing the required libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import matplotlib.colors as mcolors
```

**--AGE ANALYSIS--**

```python
df1 = pd.read_csv('../input/covid19-in-india/AgeGroupDetails.csv')
```

```python
df1.head()
```

Creating a function to plot pie charts.

```python
def plot_pie_charts(x, y, title):
    c = random.choices(list(mcolors.CSS4_COLORS.values()),k = 10)
    plt.figure(figsize=(20,15))
    plt.title(title, size=20)
    plt.pie(y, colors = c)
    plt.legend(x, loc='best', fontsize=15)
    plt.show()
```

Plotting a pie chart of the age distribution of the confirmed cases.

```python
plot_pie_charts(df1['AgeGroup'], df1['TotalCases'], 'Age distribution')
```

 We can see that most of the cases are of people 20-29 age range with the ranges 30-39,40-49,50-59,60-69 also having a high number of cases.

**--Analysis of number of public health facilities and number of beds--**

```python
df2 = pd.read_csv('../input/covid19-in-india/HospitalBedsIndia.csv')
```

```python
df2.tail()
```

```python
def plot_bar_graphs(x, y, title):
    plt.figure(figsize=(20, 12))
    plt.barh(x, y)
    plt.title(title, size=20)
    plt.xticks(size=15)
    plt.yticks(size=20)
    plt.show()
```

```python
df2 = df2.drop(df2.index[36])
```

```python
plot_bar_graphs(df2['State/UT'], df2['NumRuralHospitals_NHP18'], '# of Rural Hospitals in different regions')
```

Uttar pradesh has the most and is the only state with over 4000 rural hospitals.

```python
plot_bar_graphs(df2['State/UT'], df2['NumRuralBeds_NHP18'], '# of Rural Beds in different regions')
```

Tamil Nadu is the state with the most Beds in the rural areas with over 40000 beds.
Uttar Pradesh is a close second.

```python
plot_bar_graphs(df2['State/UT'], df2['NumUrbanHospitals_NHP18'], '# of Urban Hospitals in different regions')
```

Tamil Nadu has the most urban hospitals with over 500 of them.
Maharashtra is the only other state with over 400 urban hospitals.

```python
plot_bar_graphs(df2['State/UT'], df2['NumUrbanBeds_NHP18'], '# of Urban Beds in different regions')
```

West Bengal has the most beds in Urban areas with almost 60000 beds.

```python
plot_bar_graphs(df2['State/UT'], df2['TotalPublicHealthFacilities_HMIS'], '# of Public Health Facilities in different regions')
```

Uttar Pradesh has the most public health facilities with over 4000.

```python
 plot_bar_graphs(df2['State/UT'], df2['NumPublicBeds_HMIS'], '# of Public Beds in different regions')
```

Tamil Nadu has the most public beds with over 70000 of them.
Maharashtra and Andhra Pradesh are the only other state with over 60000 beds.
