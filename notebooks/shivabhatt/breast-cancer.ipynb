Here data is divided in to three parts first is Mean,second Standard Error and third is worst and each parametre contain 10 Paramete(rradius, texture,area, perimeter, smoothness,compactness,concavity,concave points,symmetry and fractal dimension),now we will divide the features according to their category.

```python
features_mean= list(data.columns[1:11])
features_se= list(data.columns[11:20])
features_worst=list(data.columns[21:31])
print(features_mean)
print("-----------------------------------")
print(features_se)
print("------------------------------------")
print(features_worst)
```

```python
sns.countplot(data['diagnosis'],label="Count")

#graph is showing there is more number of bengin stage of cancer than malignant.
```

```python
prediction_var = ['texture_mean','perimeter_mean','smoothness_mean','compactness_mean','symmetry_mean']
```

```python
train_df, test_df = train_test_split(data, test_size = 0.3)# in this our main data is splitted into train and test
# we can check their dimension
print(train_df.shape)
print(test_df.shape)
```

```python
X = train_df[prediction_var]                               # taking the data input 
y = train_df.diagnosis                                       # This is output of our data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=123)
```

**Training a dataset using Random forest Classifier**

```python
rf=RandomForestClassifier(n_estimators=100)              # A simple random forest model
rf.fit(X_train,y_train)
rf_prediction=rf.predict(X_test)

metrics.accuracy_score(rf_prediction,y_test)
```

```python
svc = SVC()
svc.fit(X_train,y_train)
svc_prediction=svc.predict(X_test)

metrics.accuracy_score(svc_prediction,y_test)
```
