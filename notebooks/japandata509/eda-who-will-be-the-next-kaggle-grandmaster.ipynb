<h2 style='background:#FFFFFF; border:0; color:black'><center>Do you know Kaggle Tiers?<center><h2>


![](https://www.kaggle.com/static/images/tier-animation-transparent.gif)
* [Image Source](https://www.kaggle.com/progression)

* On Kaggle, there are 5 tiers such as <span style="font-size:20px;color:green;">Novice</span>,<span style="font-size:20px;color:blue;">Contibutor</span>,<span style="font-size:20px;color:purple;">Expert</span>,<span style="font-size:20px;color:orangered;">Master</span>,and <span style="font-size:20px;color:gold;">Grandmaster</span>.

# <span style="color:green">Novice</span>

* When you just registered Kaggle, you were difinitely Novice.
* Novices can not downvote other's comment or topics, and their upvote doesn't count towards medals.

# <span style="color:blue">Contributor</span>

* To become a contributer, you have to meet some conditions below.
> * Add a bio to your profile
> * Add your location
> * Add your occupation
> * Add your organization
> * SMS verify your account
> * Run 1 script
> * Make 1 competition or task submission
> * Make 1 comment
> * Cast 1 upvote

* These conditions are very easy, so I think that Novices should become Contributor ASAP!

# <span style="color:purple">Expert</span>

* Unlike Novice and Contributer, you have to get medals to become an Expert.

| The type of Expert | Requirement |
| --- | --- |
|  Competitons Expert | 2 Bronze Medals|
| Datasets Expert | 3 Bronze Medals|
| Notebooks Expert | 5 Bronze Medals|
| Discussion Expert| 50 Bronze Medals

> 
* I will explain how to get a medal later.

# <span style="color:orangered">Master</span>

* Whether Competition Master or other Masters, becoming a Master is very difficult.
> 

| The type of Master | Requirement |
| --- | --- |
|  Competitons Master | 1 Gold Medal and 2 Silver Medals|
| Datasets Master | 1 Gold Medal and 4 Silver Medals|
| Notebooks Master | 10 Silver Medals|
| Discussion Master| 50 Silver Medals and Total 200 Medals|

# <span style="color:gold">Grandmaster</span>

* Only a very limited number of Kagglers can become a Grandmaster.

| The type of Grandmaster | Requirement |
| --- | --- |
|  Competitons Grandmaster | 5 Gold Medals and Solo Gold|
| Datasets Grandmaster | 5 Gold Medals and 5 Silver Medals|
| Notebooks Grandmaster | 15 Gold Medals|
| Discussion Grandmaster| 50 Gold Medals and Total 500 Medals|

> 
* "Solo Gold" means the gold medal won by yourself only.
* In other words, to get a solo gold, you shouldn't merge a team.

# How do we get a <span style="font-size:30px;color:gold;">medal</span>?

* To become Expert,Master,and Grandmaster, we have to get a medal.
* Then, how do we get a medal on Kaggle?

## Competition Medals

* To get a Competition medal, you need to get good results in the competition.


| | 0-99 teams | 100-249 teams | 250-999 teams| 1000+ teams |
|--- | --- | --- | --- | --- | 
|Bronze|  Top 40% | Top 40%| Top 100| Top 10%
|Silver| Top 20% | Top 20%| Top 50 | Top 5%|
|Gold| Top 10% | Top 10| Top 10 + 0.2% | Top 10 + 0,2% | 


> 
* As you can see from the table above, getting a medal in competiton is not easy.
* Only a very talented Kagglers can get a gold medal.

## Datasets/Notebooks/Discussion Medals

* To get medals in Datasets/Notebook/Discussion, you need to upload public datasets/notebooks or post a comment and get upvotes from other Kagglers.

|Datasets/Notebooks Medals | Requirement |
|--- | --- |
|Bronze|  5 Votes | 
|Silver| 20 Votes | 
|Gold| 50 Votes|

|Discussion Medals | Requirement |
|--- | --- |
|Bronze|  1 Vote | 
|Silver| 5 Votes | 
|Gold| 10 Votes|

> 
* Not all upvote count towards medals: **votes by Novices and votes on old posts and self-votes are excluded from medal calculation.** (Quoted from [here](https://www.kaggle.com/progression))

* The difinition of **"old"** is not announced, but in my experience, posts about a month or two ago do not seem to be considered **"old"**. (If anyone knows, please let me know.)

<h2 style='background:#FFFFFF; border:0; color:black'><center>EDA: Masters and Grandmasters<center><h2>

* Here goes the main part.
* Based on the Meta Kaggle datasets, we analyze the data about Kaggle <span style="font-size:20px;color:orangered;">Masters</span> and <span style="font-size:20px;color:gold;">Grandmasters</span>.

### Note: UserAchievements.csv in "Meta Kaggle" does not contain data about Datasets Grandmaster, so if I have time, I'll add data about them.

```python
tier_df_count=Users.loc[(Users['PerformanceTier']!=5)]['PerformanceTier'].value_counts().to_frame().reset_index().rename(columns={'index':'PerformanceTier','PerformanceTier':'count'})
```

```python
tier_df_count.PerformanceTier=['Novice','Contributer','Expert','Master','Grandmaster']
```

## How many Masters/Grandmasters does Kaggle have?

```python
fig = px.bar(tier_df_count.iloc[3:], x='PerformanceTier', y='count',title='The Number of Kaggle Master/Grandmaster')
fig.show()
```

* According to the graph above, it is obivious that the number of Grandmaster is very smaller than that of Master,which shows the difficulty of becoming a Grandmaster.

## Masters/Grandmasters by Type

```python
gm_bytypes = [0,0,0,0]
m_bytypes = [0,0,0,0]
for i in UserAchievements[UserAchievements["Tier"] == 4]['AchievementType']:
    if i == "Competitions":
        gm_bytypes[0] += 1
    elif i == "Discussion":
        gm_bytypes[1] += 1
    else:
        gm_bytypes[2] += 1
for i in UserAchievements[UserAchievements["Tier"] == 3]['AchievementType']:
    if i == "Competitions":
        m_bytypes[0] += 1
    elif i == "Discussion":
        m_bytypes[1] += 1
    else:
        m_bytypes[2] += 1
```

```python
gm_bytypes[3] = 9
m_bytypes[3] = 20
```

```python
GMasters = pd.DataFrame(data=gm_bytypes,columns=['Number'],index=['Competitions','Discussion','Notebooks','Datasets'])
Masters = pd.DataFrame(data=m_bytypes,columns=['Number'],index=['Competitions','Discussion','Notebooks','Datasets'])
```

```python
fig = px.bar(GMasters, x=GMasters.index, y='Number',title='Grandmasters by Type')
fig.show()
```

```python
fig = px.bar(Masters, x=Masters.index, y='Number',title='Masters by Type')
fig.show()
```

* We can see that there are a large number of Kagglers who became a Master/Grandmaster in Competitions.
* The number of Datasets Master/Grandmaster is not included the data, so we got the number from [Kaggle Rankings](https://www.kaggle.com/rankings?group=datasets&page=1&pageSize=20).

## Grandmasters by Country

* We know that there are many people from diffrent countries on Kaggle.
* Here We'd like to summarize the number of Master/Grandmaster by country.

* Unfortunately, **"Meta Kaggle"** datasets doesn't have the information about countries kagglers live in.
* Therefore, we have to the information by hand (Webscraping is also useful, but I couldn't).
* We observe the data about Grandmaster only here, because it takes a lot of time to make datasets about Master (About 1600 people).
* Datasets about Kaggle Grandmaster can be accessed [here](https://www.kaggle.com/japandata509/kaggle-grandmasters).

```python
kaggle_gm = pd.read_csv('../input/kaggle-grandmasters/Kaggle_Grandmasters.csv')
kaggle_gm
```

```python
countries = kaggle_gm["Country"].value_counts()
gm_country = pd.DataFrame({"Country":countries.keys(),"Number of GM":countries})
#countries
gm_country.iloc[:,1:].head(10)
```

```python
fig = px.bar(gm_country, x="Number of GM", y=gm_country.index,title='Kaggle Grand Masters by Country (Bar)')
fig.show()
```

```python
fig = px.pie(gm_country, values="Number of GM", names=gm_country.index, title='Kaggle Grand Masters by Country (Pie)')
fig.update_traces(textposition='inside')
fig.show()
```

* We can see that about a quarter of Kaggle Grand Masters live in United States.
* In Europe, Russia and United Kingdom stand out.
* In Asia, many Grandmasters live in China, Japan, and India.

## How long Grandmasters enjoy Kaggle?

* Kaggle was created in 2010, and more and more poeple registered year by year.
* In general, it is said that the more time you spend, the better your skills will be, but some people do very well in short a short time.
* We observe how long Kaggle Grandmasters enjoy Kaggle here.

```python
register_year = []
for i in kaggle_gm["RegisterDate"]:
    register_year.append(i[6:10])
#kaggle_gm["RegisterDate"][0][6:10]
collections.Counter(register_year)
```

```python
x = [19,29,36,19,17,30,36,36,17,3] 
y = ['2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
fig = go.Figure([go.Bar(x=y, y=x)])
fig.update_layout(title_text='Registered Year')
fig.show()
```

* Many Grandmasters started Kaggle journey more than five years ago, but some have become Grandmaster in a year or two.

## How many medals did Kaggle Grandmasters get?

* Kaggle Grandmasters are good at getting medals, so we observe the data about the number of medals they got.

## Medals (Competitions)

```python
Gm_user = UserAchievements[UserAchievements["Tier"] == 4]
Gm_user[Gm_user["AchievementType"] == "Competitions"].sort_values(by='TotalGold')
```

```python
fig = px.histogram(Gm_user[Gm_user["AchievementType"] == "Competitions"], x="TotalGold", title="How many Gold Medal did Grandmaster win? (Competitions)")
fig.show()
```

* Most of Grandmaster got 5 to 10 gold medals.
* What's amazing is that there are two Grandmasters who have won more than 30 gold medals.
* The Grandmaster who got the most gold medals in competitions is ["Giba"](https://www.kaggle.com/titericz) (55 Gold Medals)

## Medals (Notebooks)

```python
fig = px.histogram(Gm_user[Gm_user["AchievementType"] == "Scripts"], x="TotalGold", title="How many Gold Medal did Grandmaster win? (Notebooks)")
fig.show()
```

* To become a Notebooks Grandmaster, it is required to get 15 gold medals, so  most of Grandmasters got 15 to 20 gold medals.
* The Grandmaster who got the most gold medals in notebooks is ["Chris Deotte"](https://www.kaggle.com/cdeotte) (45 Gold Medals)

## Medals (Discussion)

```python
fig = px.histogram(Gm_user[Gm_user["AchievementType"] == "Discussion"], x="TotalGold", title="How many Gold Medal did Grandmaster win? (Discussion)")
fig.show()
```

* Naturally, it is very difficult to get a gold medal in discussion (10 Upvotes are required).
* Most of Grandmaster got 50 to 100 gold medals.
* The Grandmaster who got the most gold medals in discussion is ["CPMP"](https://www.kaggle.com/cpmpml) (322 Gold Medals)
