# Visualising Our Data

```python
# Finding out the amount of songs per genre represented in the top 50

plt.figure(figsize = (6,6))
q = sns.countplot(df.Genre)
q.set(xlabel = "Genre")
q.set(ylabel = "Total songs")
q.set(title = "Number of songs represented by Genre")
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

```python
# Visualising beats per minute for different genre's

plt.figure(figsize = (6,6))
sns.barplot(x = df["Genre"], y = df["Beats.Per.Minute"], palette = "rocket")
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

```python
# Visualising energy for different genre's

plt.figure(figsize = (6,6))
q2 = sns.scatterplot(x = (df["Genre"]), y = df["Energy"], palette = "rocket", alpha = 0.75)
plt.xticks(rotation = 90)
q2.set(xlabel = "Genre")
q2.set(ylabel = "Energy")
q2.set(title = "Energy by Genre Scatterplot")
plt.tight_layout()
```

```python
# Finding out the amount of each artist in the top 50

plt.figure(figsize = (6,6))
q3 = sns.countplot(df["Artist.Name"], palette = "crest")
plt.xticks(rotation = 90)
q3.set(xlabel = "Artist Name")
q3.set(ylabel = "Total Count")
plt.tight_layout()
plt.show()
```

```python
# Checking if artists usually release songs with similar or varying danceability

plt.figure(figsize = (6,6))
q4 = sns.scatterplot(x = df["Artist.Name"], y = df["Danceability"], alpha = 0.75, palette = "rocket")
plt.xticks(rotation = 90)
q4.set(xlabel = "Artist Name")
q4.set(ylabel = "Danceability")
q4.set(title = "Danceability by Artist")
plt.tight_layout()
```

```python
# Checking if artists songs have varying popularity

plt.figure(figsize = (6,6))
q5 = sns.scatterplot(x = df["Artist.Name"], y = df["Popularity"])
plt.xticks(rotation = 90)
q5.set(xlabel = "Artist Name")
q5.set(ylabel = "Popularity")
q5.set(title = "Popularity by Artist")
plt.tight_layout()
```
