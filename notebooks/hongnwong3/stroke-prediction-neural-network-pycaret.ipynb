```python
import numpy as np
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

df= pd.read_csv('../input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')
```

```python
df
```

```python
# BMI Group
df['bmi_group']=np.nan
for x in range(len(df['bmi'])):
    if df['bmi'][x] < 18.5:
        df['bmi_group'][x]='underweight'
    if df['bmi'][x] >= 18.5 and df['bmi'][x]<25:
        df['bmi_group'][x]='health'
    if df['bmi'][x]>=25 and df['bmi'][x]<30:
        df['bmi_group'][x]='overweight'
    if df['bmi'][x]> 30:
        df['bmi_group'][x]='obese'
```

```python
# Age Group
df['age_group']=np.nan
for x in range(len(df['age'])):
    if df['age'][x] <= 3:
        df['age_group'][x]='baby'
    if df['age'][x]>3 and df['age'][x]<=12:
        df['age_group'][x]='children'
    if df['age'][x]>12 and df['age'][x]<=20:
        df['age_group'][x]='adolescence'
    if df['age'][x]>20 and df['age'][x]<=40:
        df['age_group'][x]='young adult'
    if df['age'][x]>40 and df['age'][x]<=60:
        df['age_group'][x]='adult'
    if df['age'][x]>60 and df['age'][x]<=70:
        df['age_group'][x]='middle age'
    if df['age'][x]>70:
        df['age_group'][x]='senior'
```

```python
# Glucose Level Group
df['glucose_level_group']=np.nan
for x in range(len(df['avg_glucose_level'])):
    if df['avg_glucose_level'][x]<=115:
        df['glucose_level_group'][x]='normal'
    if df['avg_glucose_level'][x]>115 and df['avg_glucose_level'][x]<=180:
        df['glucose_level_group'][x]='impaired glucose'
    if df['avg_glucose_level'][x]>180:
        df['glucose_level_group'][x]='diabetes'
```

```python
df_cp=df.drop(['id','age','avg_glucose_level','bmi'],axis=1)
```

```python
y=df_cp['stroke']
x=df_cp.drop(['stroke'],axis=1)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
x['gender']=encoder.fit_transform(x['gender'])
x['ever_married']=encoder.fit_transform(x['ever_married'])
x['work_type']=encoder.fit_transform(x['work_type'])
x['Residence_type']=encoder.fit_transform(x['Residence_type'])
x['smoking_status']=encoder.fit_transform(x['smoking_status'])
x['bmi_group']=encoder.fit_transform(x['bmi_group'].astype(str))
x['age_group']=encoder.fit_transform(x['age_group'])
x['glucose_level_group']=encoder.fit_transform(x['glucose_level_group'])
```

```python
# Standardizing
scaler=StandardScaler()
X=scaler.fit_transform(x)
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
# Create Model
model=Sequential()

# Input layer Layer
model.add(Dense(32,activation='relu',input_dim=10))

# Hidden Layer
model.add(Dense(16,activation='relu'))

# Outout Layer
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# Fit model
model.fit(X_train,y_train,batch_size=5,epochs=5)
```

```python
eval_model=model.evaluate(X_train,y_train)
```

```python
eval_model
```

```python
y_pred=model.predict(X_test)
y_pred =(y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
