```python
covid19 = pd.read_csv('/kaggle/input/corona-virus-report/usa_county_wise.csv',parse_dates=True)
```

```python
covid19.info()
```

```python
covid19 = covid19.astype({
    'UID':'int32',
    'Date':'datetime64',
    'code3':'int32',
    'Confirmed':'int32',
    'Deaths':'int32',
    'Lat':'float32',
    'Long_':'float32',
    'FIPS':'float32',
})
```
