```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import spacy
nlp = spacy.load("en_core_web_sm")
fake_data = pd.read_csv('/kaggle/input/fake-and-real-news-dataset/Fake.csv')
true_data = pd.read_csv('/kaggle/input/fake-and-real-news-dataset/True.csv')
```

### Label Data with fake=0 and true=1

```python
fake_data['label'] = 0
true_data['label'] = 1
```

```python
frames = [fake_data, true_data]
data = pd.concat(frames)
```

```python
with nlp.disable_pipes():
    vectors = np.array([nlp(each.title).vector for idx, each in data.iterrows()])
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


X_train, X_test, y_train, y_test = train_test_split(vectors, data.label, test_size=0.1, random_state=1)
model = LinearSVC(random_state=1, dual=False)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, vectors, data.label, cv=5)
scores
```

### Conclusion

I only used the document vector of the titles because I think fake news tend to use exaggerated and ambiguous expressions. 
By using Linear SVC, I successfully separated fake and true news vector into two spaces that can be seen by the high accuracy of this model.
I have also conducted cross-validation to check if the model is overfitted. 
As all scores in cross-validations are above 97, the model is not overfitted and can be considered as a sufficient model for detecting fake news.
