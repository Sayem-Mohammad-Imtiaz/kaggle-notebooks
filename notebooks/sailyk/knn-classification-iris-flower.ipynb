```python
X = df.drop("species",axis =1)
y = df["species"]
```

```python
print(X)
print("Shape of X:",X.shape)
print(y)
print("Shape of y:",y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 42)
```

```python
print("Xtrain shape:",X_train.shape)
print("ytrain shape:",y_train.shape)
print("Xtest shape:",X_test.shape)
print("ytest shape:",y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh_iris_model = KNeighborsClassifier(n_neighbors=3)
neigh_iris_model.fit(X_train, y_train)
```

```python
pickle.dump(neigh_iris_model, open("/kaggle/working/iris_classification_model.sav", 'wb'))
```

```python
with open("/kaggle/working/iris_classification_model.sav", 'rb') as file:
    iris_model = pickle.load(file)
```

```python
prediction = iris_model.predict(X_test)
```

```python
print("Prediction:",prediction)
print(len(prediction))
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix 
accuracy_score(y_test,prediction) #validation accuracy
```

```python
print(" Validation Accuracy Score:",accuracy_score(y_test,prediction))
```

```python
print("Confusion Matrix:",confusion_matrix(y_test,prediction))
```

```python
#getting training accuracy
training_prediction = neigh_iris_model.predict(X_train)
print("Training accuracy :",accuracy_score(y_train,training_prediction))
```
