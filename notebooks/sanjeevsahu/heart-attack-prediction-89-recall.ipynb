```python
X = df1.drop('output', axis=1)
y = df1['output']
```

```python
Xtrain,Xtest ,ytrain,  ytest = train_test_split(X,y, test_size = 0.3, random_state = 42)
```

```python
clf = LogisticRegression(max_iter= 1000)
clf.fit(Xtrain,ytrain)
```

```python
ypred = clf.predict(Xtest)
```

```python
accuracy_score(ypred, ytest)
```

```python
print(classification_report(ytest, ypred))
```

```python
clf1 = SVC()
clf1.fit(Xtrain,ytrain)
```

```python
ypred = clf1.predict(Xtest)
```

```python
print(classification_report(ytest, ypred))
```

Let's try multiple models at once

```python
knnc = KNeighborsClassifier().fit(Xtrain,ytrain)
gnb = GaussianNB().fit(Xtrain,ytrain)
cartc = DecisionTreeClassifier(random_state=42).fit(Xtrain,ytrain)
rfc = RandomForestClassifier(random_state=42,verbose=False).fit(Xtrain,ytrain)
gbmc = GradientBoostingClassifier(verbose=False).fit(Xtrain,ytrain)
```

```python
clf_models = [knnc,gnb, cartc,rfc,gbmc]
```

```python
for model in clf_models:
    name = model.__class__.__name__
    predict = model.predict(Xtest)
    cross_val = cross_val_score(model, Xtest, ytest, cv=10, verbose = False).mean()
    print(name + ';')
    print("----"*20)
    print(cross_val)
    print(classification_report(predict, ytest))
   
```

**CONCLUSION**
As we can see that **Gaussian Naive Bayes** algorithm gives us the best precision score and recall score.
Its a heart attack prediction problem, so it's important to recognize the people who are suceptible to heart attack on current conditions(features). 

Precision and recall become very important here because they tell what % of patients who are predicted positive are actually positive. 
RECALL is the most important metric here, as it tells us about positive cases out of all the positive cases( also the FN cases).
