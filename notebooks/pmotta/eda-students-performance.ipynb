```python
df = pd.read_csv("../input/students-performance-in-exams/StudentsPerformance.csv")
df
```

Adicionando uma coluna para representar a nota total

```python
df["total"] = (df["math score"] + df["reading score"] + df["writing score"])//3
df
```
