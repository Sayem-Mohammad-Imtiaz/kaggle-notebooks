```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
```

```python
X = data.drop('Outcome', axis=1).values
y = data['Outcome'].values

#split data
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=42, stratify=y)
```

```python
neighbors = np.arange(1,9)
dt_train_akurasi =np.empty(len(neighbors))
dt_test_akurasi = np.empty(len(neighbors))

for i,k in enumerate(neighbors):
    data_knn = KNeighborsClassifier(n_neighbors=k)
    data_knn.fit(X_train, y_train)
    
    #akurasi data train
    dt_train_akurasi[i] = data_knn.score(X_train, y_train)
    #akurasi data test
    dt_test_akurasi[i] = data_knn.score(X_test, y_test)
```

```python
plt.title('K-NN Level')
plt.plot(neighbors, dt_test_akurasi, label='Teseting Akurasi')
plt.plot(neighbors, dt_train_akurasi, label='Training Akurasi')
plt.legend()
plt.xlabel('Number of neighbors')
plt.ylabel('Akurasi')
plt.show()
```

```python
#K-NN Classsifier Neighborship

data_knn = KNeighborsClassifier(n_neighbors=7)
data_knn.fit(X_train,y_train)

#Akurasi
data_knn.score(X_test,y_test)
```
