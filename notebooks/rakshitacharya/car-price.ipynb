# PrepRocessing

```python

import numpy as np # linear algebra
import pandas as pd 
df1=pd.read_csv("/kaggle/input/vehicle-dataset-from-cardekho/car data.csv")
df1.head()
```

```python
print(df1["Fuel_Type"].unique())
print(df1["Seller_Type"].unique())
print(df1["Transmission"].unique())
print(df1["Owner"].unique())
```

```python
own={3:2}
df1["Owner"]=df1["Owner"].replace(own)
```

```python
pet={"Petrol":1,
     "Diesel":2,
     "CNG":3}
sell={"Dealer":1,
     "Individual":2}
drive={"Manual":1,
      "Automatic":2}

df1["Fuel_Type"]=df1["Fuel_Type"].replace(pet)
df1["Seller_Type"]=df1["Seller_Type"].replace(sell)
df1["Transmission"]=df1["Transmission"].replace(drive)
```

```python
print(df1["Fuel_Type"].unique())
print(df1["Seller_Type"].unique())
print(df1["Transmission"].unique())
print(df1["Owner"].unique())
```

```python
df1.head()
```

```python
df1.drop(["Car_Name"],axis=1,inplace=True)
```

```python
df1["Year"]=2021-df1['Year']
```

```python
import seaborn as sns
sns.heatmap(df1.corr(),annot=True);
```

```python
sns.pairplot(df1);
```

```python
X=df1[["Year","Present_Price","Kms_Driven","Fuel_Type","Seller_Type","Transmission","Owner"]]
y=df1["Selling_Price"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
import matplotlib.pyplot as plt
model = ExtraTreesRegressor()
model.fit(X,y)
```

```python
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(5).plot(kind='barh')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor(n_estimators= 1000,
 min_samples_split= 2,
 min_samples_leaf= 1,
 max_features= 'sqrt',
 max_depth= 25)
```

```python
regressor.fit(X_train,y_train)
y_pred = regressor.predict(X_test)
```

```python

from sklearn.metrics import mean_absolute_error, mean_squared_error
import math
print(math.sqrt(mean_squared_error(y_pred, y_test)))
print(mean_absolute_error(y_pred, y_test))
```
