```python
def boxplot(query, y):
    f, ax = plt.subplots(figsize=(20, 9))
    ax = sns.boxplot(x="label", y=y, data=music.query(query))
    ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
    plt.show()
    
boxplot('label >= 1925 & label <= 1995', "TimbreAvg7")
```

```python
#group the songs by decade 
music['decade'] = music.label.apply(lambda year : year-(year%10))
```

```python
music4 = music[["decade","TimbreAvg1", "TimbreAvg2", "TimbreAvg3", "TimbreAvg4","TimbreAvg5", "TimbreAvg6"
              ,"TimbreAvg7","TimbreAvg8","TimbreAvg9","TimbreAvg10","TimbreAvg11","TimbreAvg12"]]
columns = music4.groupby(['decade']).mean().columns
labels = ["{:02d}'s".format(l%100) for l in sorted(music4.decade.unique())]
fig, ax = plt.subplots(figsize=(20,5)) 
sns.heatmap(music4.groupby(['decade']).mean().iloc[:,0:20], yticklabels=labels)
plt.ylabel("Release Decade")
plt.xlabel("Features (Mean)")
plt.show()
```

```python
music4['decade'].value_counts()
```

```python
decade_1920 = music4[(music4['decade']==1920)].sample(n=100, random_state=1)
decade_1930 = music4[(music4['decade']==1930)].sample(n=100, random_state=1)
decade_1940 = music4[(music4['decade']==1940)].sample(n=100, random_state=1)
decade_1950 = music4[(music4['decade']==1950)].sample(n=100, random_state=1)
decade_1960 = music4[(music4['decade']==1960)].sample(n=100, random_state=1)
decade_1970 = music4[(music4['decade']==1970)].sample(n=100, random_state=1)
decade_1980 = music4[(music4['decade']==1980)].sample(n=100, random_state=1)
decade_1990 = music4[(music4['decade']==1990)].sample(n=100, random_state=1)
decade_2000 = music4[(music4['decade']==2000)].sample(n=100, random_state=1)
sample = pd.concat([decade_1920,decade_1930,decade_1940,decade_1950,decade_1960,decade_1970,decade_1980,decade_1990,decade_2000])
sample.info()
```

```python
sample4 = sample[["decade","TimbreAvg1", "TimbreAvg2", "TimbreAvg3", "TimbreAvg4","TimbreAvg5", "TimbreAvg6"
              ,"TimbreAvg7","TimbreAvg8","TimbreAvg9","TimbreAvg10","TimbreAvg11","TimbreAvg12"]]
columns = sample4.groupby(['decade']).mean().columns
labels = ["{:02d}'s".format(l%100) for l in sorted(sample4.decade.unique())]
fig, ax = plt.subplots(figsize=(20,5)) 
sns.heatmap(sample4.groupby(['decade']).mean().iloc[:,0:20], yticklabels=labels)
plt.ylabel("Release Decade")
plt.xlabel("Features (Mean)")
plt.show()
```

```python
sample4['decade']=sample4['decade'].astype('str')
sample4.set_index('decade', inplace=True)
sample4.info()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaled_values = pd.DataFrame(MinMaxScaler().fit_transform(sample4)).rename(columns={0:'TimbreAvg1',1: 'TimbreAvg2',2: 'TimbreAvg3',
                                                                 3:'TimbreAvg4',4:'TimbreAvg5', 5:'TimbreAvg6',
                                                                 6:'TimbreAvg7',7:'TimbreAvg8',8:'TimbreAvg9',
                                                                 9:'TimbreAvg10', 10:'TimbreAvg11', 11:'TimbreAvg12'})
sample4.loc[:,:] = scaled_values
scaled_values.describe()
```
