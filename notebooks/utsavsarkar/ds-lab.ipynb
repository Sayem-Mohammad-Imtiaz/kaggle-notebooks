```python
import pandas as pd
```

```python
dataset = pd.read_csv('../input/indian-food-101/indian_food.csv')
dataset.head()
```

```python
dataset.tail()
```

```python
X = dataset.iloc[:,2].values
```

Q1. percentage of veg and non veg

```python

veg = 0
nveg = 0
for i in X:
    if i == 'vegetarian':
        veg += 1
    else:
        nveg += 1
print(f"Percentage of veg is {veg/len(X)*100}%")
print(f"Percentage of non veg is {nveg/len(X)*100}%")
```

```python
import matplotlib.pyplot as plt
plt.hist(X, bins = 3)
plt.show()
```

Q2. Cook time and preparation time plot

```python
cook_time = dataset.iloc[:, 4].values
prep_time = dataset.iloc[:, 3].values
```

```python
plt.scatter(cook_time,prep_time)
plt.title("Overall Time")
plt.xlabel('cooking time')
plt.ylabel('prep time')
plt.show()
```

Individual plot

```python
veg_prep_time = []
nveg_prep_time = []
for i in range(len(X)):
    if(X[i]=='vegetarian'):
        veg_prep_time.append(prep_time[i])
    else:
        nveg_prep_time.append(prep_time[i])

veg_cook_time = []
nveg_cook_time = []
for i in range(len(X)):
    if(X[i]=='vegetarian'):
        veg_cook_time.append(cook_time[i])
    else:
        nveg_cook_time.append(cook_time[i])
```

```python
plt.scatter(veg_cook_time,veg_prep_time,color='green')
plt.scatter(nveg_cook_time,nveg_prep_time,color='yellow')
plt.xlabel('cooking time')
plt.ylabel('prep time')
plt.show()
```
