```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data_test = pd.read_csv('../input/us-census-data/adult-test.csv', names=columns, skiprows=[0],
                        skipinitialspace=True, na_values="?")
data_test = pd.DataFrame(data_test)

data_test.head()
```

```python
data_train = pd.read_csv('../input/us-census-data/adult-training.csv', names=columns, 
                            skipinitialspace=True, na_values="?")
data_train = pd.DataFrame(data_train)

data_train.head()
```

```python
columns = ['age', 'employment', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation',
          'family', 'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country',
          'income_class']
```

```python
data_train.shape, data_test.shape
```

```python
data = data_train.append(data_test)
```

```python
data.describe()
```

```python
data.info()
```

```python
data = data.dropna()
```

```python
data.age.value_counts()
```

```python
data.drop('fnlwgt', axis=1, inplace=True)
```

```python
data.head()
```

```python
data.income_class.value_counts()
```

```python
data.income_class = data.income_class.map({"<=50K":0, "<=50K.":0, ">50K":1, ">50K.":1})
y = data.income_class
```

```python
pd.crosstab(data.education, data.education_num)
```

```python
data.drop('education', axis=1, inplace=True)
```

```python
plt.scatter(range(1,17), y.groupby(data.education_num).mean())
plt.show()
```

```python
data.education_num[data.education_num < 8] = 8
```

```python
plt.scatter(range(8,17), y.groupby(data.education_num).mean())
plt.show()
```

```python
data.sex = data.sex.map({"Female":1, "Male":0})
```

```python
data.head()
```

```python
data.employment.value_counts()
```

```python
data.employment.value_counts()
```

```python
data.employment[data.employment == "Without-pay"] = "Self-emp-not-inc"
```

```python
data.marital_status.value_counts()
```

```python
data.marital_status[data.marital_status == "Married-spouse-absent"] = "Married-spouse"
data.marital_status[data.marital_status == "Married-AF-spouse"] = "Married-spouse"
data.marital_status[data.marital_status == "Married-civ-spouse"] = "Married-spouse"
```

```python
data.occupation.value_counts()
```

```python
data.occupation[data.occupation == "Armed-Forces"] = "Protective-serv"
```

```python
data.head()
```

```python
data.family.value_counts()
```

```python
data.race.value_counts()
```

```python
data["capital_gain"] = np.log1p(data.capital_gain)
data["capital_loss"] = np.log1p(data.capital_loss)
```

```python
data.hours_per_week.value_counts()
```

```python
plt.hist(data.hours_per_week)
plt.xlims = "Hours per week"
plt.show()
```

```python
data.native_country = (data.native_country == "United States").astype(int)
```
