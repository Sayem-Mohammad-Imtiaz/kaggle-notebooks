## CDF

```python
counts,bin_edges=np.histogram(df['axil_nodes'],bins=20,density = True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.xlabel('axil_nodes')
plt.show()



```

```python
sns.boxplot(x='surv_status', y= 'axil_nodes' , data =df)
plt.show()
```

This graph shows that more number of axil nodes the more patient likely die.


```python
sns.violinplot(x='surv_status',y='axil_nodes', data=df,size=9)
plt.show()
```

after observing above plot we can see that, the large number of patients who survived had 0 axil nodes or doesn't had it all
patients and who died had axil node greater than 1 and as the concentration of axil node increases the repective patient more likely to die.
