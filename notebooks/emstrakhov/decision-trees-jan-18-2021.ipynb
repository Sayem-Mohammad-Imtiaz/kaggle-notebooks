```python
df = pd.read_csv('/kaggle/input/cardiovascular-disease-dataset/cardio_train.csv', 
                 sep=';',
                 index_col=0)
df.head()
```

```python
df['age_years'] = df['age'] / 365.25
df_1 = df.drop('age', axis=1)
df_1.head()
```

```python
X = df_1.drop('cardio', axis=1)
y = df_1['cardio']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=18)
```

```python
from sklearn.tree import DecisionTreeClassifier
my_tree = DecisionTreeClassifier()

my_tree.fit(X_train, y_train)
y_pred = my_tree.predict(X_valid)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_valid, y_pred))
```

```python
print(my_tree.get_depth())
```

```python
my_tree_1 = DecisionTreeClassifier(max_depth=5)

my_tree_1.fit(X_train, y_train)
y_pred = my_tree_1.predict(X_valid)

from sklearn.metrics import accuracy_score
print(accuracy_score(y_valid, y_pred))
```
