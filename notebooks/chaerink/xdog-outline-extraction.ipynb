# CV2

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def dog(img, size=(0,0), k=1.6, sigma=0.5, gamma=1):
    img1 = cv2.GaussianBlur(img, size, sigma)
    img2 = cv2.GaussianBlur(img, size, k*sigma)
    return img1-gamma*img2

def xdog(img, k=1.6, sigma=0.5, gamma=1, epsilon=1, phi=1):
    aux = dog(img, k=k, sigma=sigma, gamma=gamma) / 255.0
    for i in range(0, aux.shape[0]):
        for j in range(0, aux.shape[1]):
            if aux[i,j] < epsilon:
                aux[i,j] = 255
            else:
                aux[i,j] = 255 * (1+np.tanh(phi*aux[i,j]))
    return aux

def edge_dog(img, sigma=0.5, k=200, gamma=0.98):
    aux = dog(img, sigma=sigma, k=k, gamma=gamma)
    for i in range(0,aux.shape[0]):
        for j in range(0,aux.shape[1]):
            if(aux[i,j] > 0):
                aux[i,j] = 255
            else:
                aux[i,j] = 0
    return aux
```

```python
ims = [cv2.imread("../input/celeba-dataset/img_align_celeba/img_align_celeba/00000{}.jpg".format(i)) for i in range(1,6)]
ims = [cv2.cvtColor(img, cv2.COLOR_BGR2RGB) for img in ims] 
fig, ax = plt.subplots(1,5, figsize=(13,5))
for i in range(5):
    ax[i].imshow(ims[i])
    ax[i].axis("off")
```

```python
imgs = [cv2.imread("../input/celeba-dataset/img_align_celeba/img_align_celeba/00000{}.jpg".format(i)) for i in range(1,6)]
imgs = [cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) for img in imgs] 
fig, ax = plt.subplots(1,5, figsize=(13,5))
for i in range(5):
    ax[i].imshow(imgs[i])
    ax[i].axis("off")
```

```python
def xdog_plot(imgs, eps=1, sigma=1, phi=1):
    fig, ax = plt.subplots(1, len(imgs), figsize=(13,5))
    for i in range(len(imgs)):
        ax[i].imshow(xdog(imgs[i], epsilon=eps, sigma=sigma, phi=phi), cmap="Greys")
        ax[i].axis("off")
        
def edge_dog_plot(imgs, sigma=1, k=1, gamma=1):
    fig, ax = plt.subplots(1, len(imgs), figsize=(13,5))
    for i in range(len(imgs)):
        ax[i].imshow(edge_dog(imgs[i], gamma=gamma, sigma=sigma, k=k), cmap="Greys")
        ax[i].axis("off")
```
