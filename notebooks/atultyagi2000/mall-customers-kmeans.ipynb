```python
inVout=data.iloc[:,[3,4]]
inVout
```

```python
plt.scatter(inVout.iloc[:,0],inVout.iloc[:,1])
```

**KMeans  cluster**

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=5)
kmeans.fit(np.array(inVout))
```

```python
plt.scatter(inVout.iloc[:,0],inVout.iloc[:,1], c=kmeans.labels_, cmap='rainbow')
plt.show()
```

```python
inVout[kmeans.labels_==3].mean()
```

```python
plt.scatter(inVout[kmeans.labels_==0].iloc[:,0],inVout[kmeans.labels_==0].iloc[:,1],label='zero')
plt.scatter(inVout[kmeans.labels_==1].iloc[:,0],inVout[kmeans.labels_==1].iloc[:,1],label='one')
plt.scatter(inVout[kmeans.labels_==2].iloc[:,0],inVout[kmeans.labels_==2].iloc[:,1],label='two')
plt.scatter(inVout[kmeans.labels_==3].iloc[:,0],inVout[kmeans.labels_==3].iloc[:,1],label='three')
plt.scatter(inVout[kmeans.labels_==4].iloc[:,0],inVout[kmeans.labels_==4].iloc[:,1],label='four')
plt.title("Clusters of Clients")
plt.xlabel("Annual Income ($K)")
plt.ylabel("Spending Score (0-100)")
plt.legend()
plt.show()
```

From the above sctter you can clearely see that which cluster of customers is benificial and target for Mall and which is not the Target to improve profits
