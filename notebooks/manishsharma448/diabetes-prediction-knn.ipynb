```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("../input/diabetes-data-set/diabetes.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
y=df["Outcome"]
```

```python
X=df.drop("Outcome",axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
y.replace(to_replace =[1], value ="Diabetic",inplace=True)
```

```python
y.replace(to_replace =[0], value ="Not Diabetic",inplace=True)
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1,stratify=y)
```

```python
X_train.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
type(X_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier(n_neighbors=9,metric="euclidean")
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
X_pred=df.drop("Outcome",axis=1)
```

```python
X_pred=scaler.transform(X_pred)
```
