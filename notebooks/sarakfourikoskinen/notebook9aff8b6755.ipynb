```python
#check if have rows empty complety
def count_nan_rows(column_1 , column_2, column_3, column_4):
    # check Biggest row of NAN
    list_triple_nan =[]
    list_fourth_nan =[]
    for x in column_1:
        for y in column_2:
            for n in column_3:
                for p in column_4:
                    if x == y and x == n and p == x:
                        list_fourth_nan.append(x)
                    elif x== y and x == n:
                        list_triple_nan.append(x)
                    elif x==p and x == n:
                        list_triple_nan.append(x)
                    elif p == n and y == p:
                        list_triple_nan.append(x)
    return  list_triple_nan, list_fourth_nan


                   
    
            
```
