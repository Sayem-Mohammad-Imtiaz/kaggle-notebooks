```python
data['test preparation course'].replace("none", np.NaN, inplace=True)
```

```python
msno.bar(data)
```

```python
# Filling Nan values with "not completed"
values = {"test preparation course":"not completed"}
data = data.fillna(value=values)
```

```python
sns.pairplot(data)
```

```python
test_preparation_cours = {"completed":0, "not completed":1}
parental_level_of_education = {"bachelor's degree": 4, 'some college':3 , "master's degree": 5, "associate's degree":6 ,'high school':2 , 'some high school':1}
```

```python
race_ethnicity = {'group B':2, 'group C':3, 'group A':1, 'group D':4, 'group E':5}
gender = {'female':1, 'male':0}
lunch = {'standard':0, 'free/reduced':1}
```

```python
data['gender'] = data['gender'].map(gender)
data['test preparation course'] = data['test preparation course'].map(test_preparation_cours)
data['race/ethnicity'] = data['race/ethnicity'].map(race_ethnicity)
data['parental level of education'] = data['parental level of education'].map(parental_level_of_education)
data['lunch'] = data['lunch'].map(lunch)
```
