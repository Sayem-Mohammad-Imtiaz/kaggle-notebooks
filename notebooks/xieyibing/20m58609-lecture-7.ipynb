 # **Global earthquake map**

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df1=pd.read_csv('../input/202105-recent-earthquake/query.csv',header=0)
df1.index=pd.to_datetime(df1['time'])


df1['time']=df1.index.strftime('%Y-%m-%d %H:00:00')
df1=df1.sort_index()
print(df1.columns)
fig=px.scatter_geo(df1,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='RdBu',range_color=(0.0,5.0))
fig.update_layout(title_text='global earthquake map',title_x=0.5)
fig.show()
```

On 25 May, following the eruption of Mount Nyiragongo volcano, A series of earthquakes have hit western Rwanda's Rubavu district,19km northeast of Gisenyi.

According to the US's report, magnitude of the earthquake reached 4.7.

Houses, schools, and roads were damaged. In Gisenyi, companies, markets, banks and etc. were closed. People's daily life was heavily influenced. In neighboring Goma, the earthquake flattened several buildings, and three people were injured by falling debris.

It was the strongest earthquake in a swarm related to the 2021 eruption of Mount Nyiragongo in the DRC.

# **People fully vaccinated per hundred**

```python
df=pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df.index=pd.to_datetime(df['date'])
df=df.sort_index()
print(df.columns)

fig=px.choropleth(df,locations='country',locationmode='country names',color='people_fully_vaccinated_per_hundred',animation_frame='date',hover_name='country',color_continuous_scale='bugn',range_color=(0.0,50.00))
fig.update_layout(title_text='people fully vaccinated per hundred',title_x=0.5)
fig.show()

plotly.offline.plot(fig,filename='vaccinations.html')
```

I visualized the global ratio of people fully vaccinated per hundred per country. According to the map, judging from shades of green, a preliminary understanding of the vaccination situation in different countries can be known. Then we can learn from it how much a country has done to prevent and treat the COVID-19 and the citizens' resistance to this virus from the perspective of vaccination.
