```python
plt.figure(figsize=(9, 8))
sns.distplot(orders['order_number'], color='g', bins=50)
plt.show()
#kebanyakan orang ordernya dikit-dikit
```

```python
plt.figure(figsize=(14,8))
sns.countplot(x="days_since_prior_order", data=orders, color='blue')
plt.xlabel('Days since prior order')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(14,8))
sns.countplot(x="order_dow", data=orders)
plt.xlabel('day of week')
plt.ylabel('Count')
plt.show()
#order lebih banyak terjadi di hari ke- dan ke-1
```

```python
plt.figure(figsize=(14,8))
sns.countplot(x="order_hour_of_day", data=orders)
plt.xlabel('Hour of day')
plt.ylabel('Count')
plt.show()
#biasanya orang-orang beli pas siang hari (jam 9 sampai 15)
```

```python
order_group_waktu = orders.groupby(["order_dow", "order_hour_of_day"])["order_number"].aggregate("count").reset_index()
order_group_waktu.head(30)
```

```python
order_group_waktu = order_group_waktu.pivot('order_dow', 'order_hour_of_day', 'order_number')
order_group_waktu.head()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(order_group_waktu, cmap='ocean')
plt.show()
```

```python
op_prior = pd.merge(op_prior, products, on='product_id', how='inner')
op_prior = pd.merge(op_prior, aisles, on='aisle_id', how='inner')
op_prior = pd.merge(op_prior, departments, on='department_id', how='inner')
```

```python
op_prior.head()
```

```python
penjualan_produk = op_prior['product_name'].value_counts().reset_index()
penjualan_produk.columns =['product_name', 'freq']
penjualan_produk.head(30)
#barang dengan order terbanyak adalah pisang
```

```python
aisle_diminati = op_prior['aisle'].value_counts().reset_index()
aisle_diminati.columns =['aisle', 'freq']
aisle_diminati.head(30)
```

```python
department_diminati = op_prior['department'].value_counts().reset_index()
department_diminati.columns =['department', 'freq']
department_diminati.head(30)
#dari sini ditampilkan department dengan penjualan terbanyak
```
