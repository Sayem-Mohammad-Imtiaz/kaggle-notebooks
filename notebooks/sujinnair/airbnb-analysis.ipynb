```python
#importing data.
raw_dat = pd.read_csv('/kaggle/input/us-airbnb-open-data/AB_US_2020.csv')
raw_dat.head(3)
```

```python
# a summary of the data.
raw_dat.describe(include = "all")
```

```python
# checking for unusable values.
print(pd.isnull(raw_dat).sum())
```

On observation it appears that the feature 'neighbourhood_group' has a lot of missing values and seems obselete, so removing that feature.

```python
#dropping the feature neighbourhood_group.
del raw_dat['neighbourhood_group']
```

Now there were other features too with missing values like, 'reviews_per_month' and 'last_review'. To do further analysis let's do some cleaning first.

```python
#distribution of missing values.
missing = sns.heatmap(pd.isnull(raw_dat).T)
missing.set_title('Missing Values Proportion',fontweight='bold')
```

Last review date is not much of a feature useful for prediction.

```python
#dropping columns that are not necessary
del raw_dat['last_review']
del raw_dat['id']
del raw_dat['host_id']
del raw_dat['host_name']
del raw_dat['neighbourhood']
del raw_dat['room_type']
```

```python
#dropping the rows and colunns with null values
raw_dat = raw_dat.dropna()
```

Let's detect outliers, but first we need to get all the numeric features where we can detect for outliers.

```python
#collecting numeric features
numeric_features = raw_dat.select_dtypes(include = ['int64','float64']).columns
non_numeric = raw_dat.select_dtypes(include = ['object']).columns
```

Let's do a 'numeric_features' distribution to view the feature values before outlier removal.

```python
# density distribution of numerical features
figs, axes = plt.subplots(nrows=2, ncols=4)
ind = 0
figs.set_figheight(17)
figs.set_figwidth(25)
for row in axes:
    for col in row:
        raw_dat[numeric_features[ind]].plot(kind='kde',ax=col)
        col.set_title(numeric_features[ind] +' Distribution',fontsize=16,fontweight='bold')
        ind += 1
```
