# Import required Library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Get Info of Dataset

```python
data = pd.read_csv('../input/goodreadsbooks/books.csv',error_bad_lines=False)
data.info()
```

# Find highest reviewed Book 

```python
reviews = data.groupby('title', as_index=False)['text_reviews_count'].sum().sort_values('text_reviews_count', ascending=False).head(10)
reviews
```

# Show result graphically

```python
plt.figure(figsize=(10,5))
plt.title("Most Text Reviwed Books")
sns.barplot(y='title', x='text_reviews_count', data=reviews);
```

# **Result**

The above visualization show that the book "**Twilight (Twilight # 1)**" got highest text reviews
