Let's check correlations within the data

```python
pok_cor=pok.corr()
print(pok_cor)
```

**1. what does a negative value mean?** -it means that one variable increases as the other decreases

**2. what does a positive value mean?**  -it means that  both variables move in the same direction (increase or decrease)

**3. what does a higher number mean versus a lower number?** -  The closer the value of correlation is to +1, the stronger the linear relationship between variables. So higher number if it's positive means that if one variable change, let's say decrease it will cause linear decrease of the other as well.

```python
sn.heatmap(pok_cor, 
        xticklabels=pok_cor.columns,
        yticklabels=pok_cor.columns)
```

We can **strong corelation** between Total and each strength of a pokemon, as each of them contribute to total.

There is **moderate correlation** between:

1) Defense and Sp.def (damage resistance against special attacks). It means if a pokemon is srong at defense there are good chances that he will be strong at special defense and verso versa.

2)Sp.def and Sp. Attack (special attack).  It means if a pokemon is srong at special defense there are good chances that he will be strong at special attack and verso versa.

3) Speed and Sp. Attack .  It means if a pokemon is srong at speed there are good chances that he will be strong at special attack and verso versa.

There is **no correlation** between Generation and any of the strengths of pokemons.
