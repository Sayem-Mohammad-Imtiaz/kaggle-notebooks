```python
train_df = pd.read_csv('/kaggle/input/mobile-price-classification/train.csv')
test_df = pd.read_csv('/kaggle/input/mobile-price-classification/test.csv')
```

```python
from sklearn.model_selection import train_test_split
train_df.head()
```

```python
X=train_df.drop(columns=['price_range'])
y=train_df['price_range']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
knn = KNeighborsClassifier()

from sklearn.model_selection import GridSearchCV
parameters = {'n_neighbors':[1,100],
             'p':[1,4]}
clf = GridSearchCV(knn, parameters)
clf = clf.fit(X_train, y_train)
clf.best_estimator_
```

```python
knn = knn.fit(X_train,y_train)
y_predict = knn.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict)
```
