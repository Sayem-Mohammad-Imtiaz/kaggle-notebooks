```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        file = os.path.join(dirname, filename)
        print(file)
```

```python
# Load CSV
df = pd.read_csv(file)
df.head()
```

```python
# Convert "Gender" from categorical to int
le = LabelEncoder()
df['Gender'] = le.fit_transform(df['Gender'])
df.head()
```

```python
# Scale data: this is important because we are measuring distances
scaler = MinMaxScaler()

x = df.values # df to np.array
x_scaled = scaler.fit_transform(x)

df = pd.DataFrame(x_scaled)  # np. array to df
df.head()
```

```python
# Calculate SSE to find best K (elbow point) 
sse = []
max_k = 10
for k in range(1, max_k):
    k_means = KMeans(n_clusters=k)
    k_means.fit(df)
    sse.append(k_means.inertia_)
    
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(range(1, max_k),sse)

# The best k = 2
```

```python
# Cluster with k = 2 (fit + predict)
k_means2 = KMeans(n_clusters=2)
y_pred = k_means2.fit_predict(df)

df['cluster'] = y_pred
df.head()
```

```python
# These are the final centroids
k_means2.cluster_centers_
```
