```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
student = pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
student.head()
```

```python
columns1={'gender':'gender','race/ethnicity':'race','parental level of education':'parent_ed_level'
          ,'lunch':'lunch_type','test preparation course':'test_prep','math score':'math'
          ,'reading score':'reading','writing score':'writing'}
student.rename(columns=columns1,inplace=True)
student
```

```python
sns.heatmap(student.isnull())
plt.show()
print(student.shape, "\n")

plt.show()
```

```python
print('row :', student.shape[0])
print('column :', student.shape[1])
print(student.info(),'\n')
```

```python
student['total']=student['math']+student['reading']+student['writing']
student['avg']= (student['math']+student['reading']+student['writing'])/3
student
```

# basic information

this because parent education level 'high school'and 'some high school' represent same level education. So we can make thats
value as same value

```python
student['parent_ed_level']=student['parent_ed_level'].replace(['some high school'],'high school')
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,5))
basic_list=['gender','lunch_type','test_prep']
title = ['Gender','Launch Type', 'Test Preparation']
x = 0
for col in basic_list:
    sns.set_style("dark")
    ax[x].set_title(title[x])
    sns.color_palette("bright")
    sns.countplot(x=col,data=student, ax=ax[x])
    x = x+1
```

```python
fig, ax = plt.subplots(2,1, figsize=(10,16))
basic1_list=['parent_ed_level','race']
title = ['Parent Education', 'Race']
x = 0
for col in basic1_list:
    sns.set_style("dark")
    ax[x].set_title(title[x])
    sns.countplot(x=col,data=student, ax=ax[x], hue=student['gender'], 
                  order = student[basic1_list[x]].value_counts().index)
    x = x+1
```

for these information we can know about data distribution for every table with gender as bonus variable

# ADVANCE INFORMATION

after get some basic information. next we can go ahead to advanced information we can get. avg score from math,
writing and reading based on every variable(tables)

```python
student
```

```python
racescore = student.groupby('race').mean().sort_values(by='avg', ascending=False)
racescore = racescore.drop(columns ='total')
fig, ax = plt.subplots(1,2, figsize=(15,5))
sns.heatmap(racescore, ax=ax[0])
# diverging_colors = sns.color_palette("hls", 8)
# sns.set_palette(diverging_colors)
sns.barplot(x=racescore['avg'], y=racescore.index, ax=ax[1])
```

```python
genderscore = student.groupby('gender').mean().sort_values(by='avg', ascending=False)
genderscore = genderscore.drop(columns ='total')
genderscore

fig, ax = plt.subplots(2,2, figsize=(10,10))
sns.barplot(x=genderscore.index, y=genderscore['avg'], ax=ax[0,0])
ax[0,0].set_title('Average')
sns.barplot(x=genderscore.index, y=genderscore['math'], ax=ax[0,1])
sns.barplot(x=genderscore.index, y=genderscore['writing'], ax=ax[1,0])
sns.barplot(x=genderscore.index, y=genderscore['reading'], ax=ax[1,1])
```

we can see female student have better average score

```python
genderscore = student.groupby('gender').mean().sort_values(by='avg', ascending=False)
genderscore = genderscore.drop(columns ='total')
genderscore

fig, ax = plt.subplots(2,2, figsize=(10,10))
sns.barplot(x=genderscore.index, y=genderscore['avg'], ax=ax[0,0])
ax[0,0].set_title('Average Score')
sns.barplot(x=genderscore.index, y=genderscore['math'], ax=ax[0,1])
ax[0,1].set_title('Math Score')
sns.barplot(x=genderscore.index, y=genderscore['writing'], ax=ax[1,0])
ax[1,0].set_title('Writing Score')
sns.barplot(x=genderscore.index, y=genderscore['reading'], ax=ax[1,1])
ax[1,1].set_title('Reading Score')
```

# top 100 student information

we want to know distribution of top 100 student

```python
rank = student.sort_values(by='avg', ascending=False).head(100)
rank
```

```python
fig, ax = plt.subplots(1,3, figsize=(15,5))
basic_list=['gender','lunch_type','test_prep']
title = ['Gender','Launch Type', 'Test Preparation']
x = 0
for col in basic_list:
    sns.set_style("dark")
    ax[x].set_title(title[x])
    sns.color_palette("bright")
    sns.countplot(x=col,data=rank, ax=ax[x])
    x = x+1
```

```python
fig, ax = plt.subplots(1,2, figsize=(15,5))
sns.countplot(x=rank['race'],data=rank, order=rank['race'].value_counts().index, ax=ax[0])
ax[0].set_title('Group Count for 100 top student')
sns.countplot(x=rank['race'],data=rank, hue='gender' ,order=rank['race'].value_counts().index, ax=ax[1])
ax[1].set_title('Group Count for 100 top student (Gender)')
```
