### 5. Feature Engineering

```python
### feature scaling 
from sklearn.pipeline import Pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
```

```python
modelpipeline = Pipeline([('scaler',MinMaxScaler()),('model',KNeighborsClassifier())])
```

### 6. Model fitting : Binary Classifier (quality >= 7 wine = Good, quality < 7 wine = Bad)

```python
from sklearn.model_selection import cross_val_score,train_test_split
```

```python
WineData['quality2'] = ['Good' if x >= 7 else 'Bad' for x in WineData.quality]
```

```python
X_train,X_test,y_train,y_test = train_test_split(WineData.drop(['quality','quality2'],axis=1),WineData.quality2,test_size = 0.1,random_state = 1)
```

```python
X_train.shape
```

```python
cross_val_score(modelpipeline,X_train,y_train,cv=5)
```

```python
modelpipeline.fit(X_train,y_train)
```

```python
modelpipeline.score(X_train,y_train)
```
