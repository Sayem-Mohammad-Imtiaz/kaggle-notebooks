**DATA CLEANING AND PROCESSING **

```python
#Discarding the datasets ( NAN/null valaues and correlations< 0.4)

flights=flights.drop(['YEAR','FLIGHT_NUMBER','AIRLINE','DISTANCE','TAIL_NUMBER','TAXI_OUT',
                                              'SCHEDULED_TIME','DEPARTURE_TIME','WHEELS_OFF','ELAPSED_TIME',
                                              'AIR_TIME','WHEELS_ON','DAY_OF_WEEK','TAXI_IN','CANCELLATION_REASON'],
                                             axis=1)
```

```python

#check any null values
flights.info()
```

```python
#filling NAN/null_value with help of mean and fillna method

flights = flights.fillna(flights.mean())

#check if there any null values 
flights.info()
```

**# HOW MANY FLIGHTS GOT DELAYED MORE THAN 15_MIN OF "ARRIVAL_TIME"**

```python
#creating a  function "RESULT"
result=[]

for i in flights['ARRIVAL_DELAY']:
  if i > 15:
    result.append(1)
  else:
    result.append(0)
```

```python
flights['result'] = result

#count the no of flights that got delayed
flights.value_counts('result')
 
```
