```python
from sklearn.impute import SimpleImputer
imputer =SimpleImputer(strategy='median')
```

```python
df2 = df.copy()
```

```python
ft_data = df2.drop('ocean_proximity',axis=1)
```

```python
imputer.fit(ft_data)
```

```python
imputer.statistics_
```

```python
ft_data.total_bedrooms.median()
```

```python
X = imputer.transform(ft_data)
```

```python
#replace the missing values in total_bedrooms with the median values
ft_transformed = pd.DataFrame(X,columns=ft_data.columns)
ft_transformed.tail()
```
