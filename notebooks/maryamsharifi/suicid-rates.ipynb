Sucides during the years

```python
df.plot(kind='scatter',x='year',y='suicides_no',alpha=0.5,color='red',
        figsize=(10,10),
        grid=True,label='Suicides No - Year')
plt.xlabel('year')
plt.ylabel('sucides no')
plt.legend()
plt.title('Suicides during the years')
plt.show()
```

Suicid number based on country.

```python
df.plot(kind='line',x='country',y='suicides_no',alpha=0.8,color='blue',
        figsize=(10,10),
        grid=True,label='Suicides No - Country')
plt.xlabel('country')
plt.ylabel('sucides no')
plt.legend()
plt.title('Suicid number based on country')
plt.show()
```

gender and suicides number.

```python
df.plot(kind='scatter',x='sex',y='suicides_no',alpha=0.8,color='green',
        figsize=(10,10),
        grid=True,label='Suicides No - Sex')
plt.xlabel('sex')
plt.ylabel('sucides no')
plt.legend()
plt.title('Suicid number based Sex')
plt.show()
```
