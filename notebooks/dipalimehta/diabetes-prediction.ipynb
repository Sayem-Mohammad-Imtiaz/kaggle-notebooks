```python
X=df_diabetes.drop(labels='Outcome',axis=1)
y=df_diabetes['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.impute import SimpleImputer
fill_values=SimpleImputer(missing_values=0, strategy='mean')
X_train=fill_values.fit_transform(X_train)
X_test=fill_values.fit_transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
pred=rf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'bootstrap': [True],
    'max_depth': [80, 90, 100, 110],
    'max_features': [2, 3],
    'min_samples_leaf': [3, 4, 5],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [100, 200, 300, 1000]
}
# Create a based model
rf = RandomForestClassifier()
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
```

```python
grid_search.fit(X_train, y_train)
grid_search.best_params_
```

```python
rfc1=RandomForestClassifier(bootstrap=True,max_depth=90,max_features=2,min_samples_leaf=4,min_samples_split= 12,n_estimators= 300)
```

```python
rfc1.fit(X_train, y_train)
```

```python
pred_gcv=rfc1.predict(X_test)
```

```python
print(classification_report(y_test,pred_gcv))
```
