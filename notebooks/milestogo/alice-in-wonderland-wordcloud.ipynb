**Alice In Wonderland****** **Word Cloud**

Word cloud works in a simple and effective way. The number of times a word appears in a text, it gets emboldened in the image.



```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt


from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
from PIL import Image

print ('Wordcloud is installed and imported!')

import warnings 
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
alice_novel= open('../input/alice-in-wonder-land/alice_in_wonderland.txt','r').read()

print('File downloaded and saved')
```

```python
stopwords = set(STOPWORDS) #set is used to remove any redundant or useless stopwords
```

```python
alice_wc = WordCloud(
    background_color='white',
    max_words=3000,
    stopwords=stopwords
)

# generate the word cloud
alice_wc.generate(alice_novel)
```

  a word cloud object is introduced.Amazing!Initially, to keep things simple we have made a word cloud with only 3000 words.


```python
#lets visualize this now
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

So now we know that **Alice , Said , little, Queen** are some of the words frequently used. We know that said is a generic word so we will try to eliminate it

```python

stopwords.add('said') # add the words said to stopwords

# re-generate the word cloud
alice_wc.generate(alice_novel)

# display the cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopwords.add('said') # add the words said to stopwords

# re-generate the word cloud
alice_wc.generate(alice_novel)

# display the cloud
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
alice_mask = np.array(Image.open('../input/twitter/twitter.jpg')) #creating a mask on twitter image

print('File downloaded and saved')
```

```python
fig = plt.figure()
fig.set_figwidth(10)
fig.set_figheight(15)
plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear') 
plt.axis('off')
plt.show()
#Bilinear Interpolation : is a resampling method that uses the distanceweighted average of the four nearest pixel values 
#to estimate a new pixel value.
```

```python
alice_wc= WordCloud(background_color='white',max_words = 3000,stopwords=stopwords, mask = alice_mask)
alice_wc.generate(alice_novel)
fig=plt.figure()
fig.set_figwidth(10)
fig.set_figheight(15)
plt.axis('off')
plt.imshow(alice_wc, interpolation='bilinear')
plt.show()
```
