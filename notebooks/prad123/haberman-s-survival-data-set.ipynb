There are 306 observations with no missing values. 

The dataset contains the following fields -
* Age : Age of the patient at the time of operation
* Operation_year: Year on which the operation was conducted. (Between 1958-1970)
* Axil_nodes: Number of tumors
* Survival_status: Whether the patient survived for more than 5 years post surgery

**The objective of the analysis:** given the dataset classify patients chance of survival for 5+ years from the date of surgery.

```python
haberman["Survival_status"].value_counts()
```

```python
haberman["Survival_status"].value_counts(normalize=True)
```

Imbalanced dataset with 74% of observation pertaining to patients who survived for 5+ years post surgery.

```python
sns.set_style("whitegrid");
sns.pairplot(haberman, hue="Survival_status", size=4)
plt.legend
```

* No possible linear relationship between variables found. 
* Simple classification using "lines" and "if-else" is not possible.

```python
sns.FacetGrid(haberman, hue="Survival_status", size=5) \
   .map(sns.distplot, "Age") \
   .add_legend();
```

No visible pattern on survival vs. age 

```python
sns.FacetGrid(haberman, hue="Survival_status", size=5) \
   .map(sns.distplot, "Axil_nodes") \
   .add_legend();
```

Plausible with a high missclassification rate. Looks like chances of survival is high when the number of Axil nodes are <= 2. Let's analyze this further...

```python
sns.distplot(haberman["Axil_nodes"], kde_kws=dict(cumulative=True))
```

 With a threshold of 2, the miss-classification rate for survival based on Axil nodes will be as high as 50% (as good as tossing a coin).

```python
sns.boxplot(x='Survival_status',y='Axil_nodes', data=haberman)
```

Same observation as above. Classification rate will be too high.

**Conclusion:**

Given variables are not sufficient to find a pattern in the dataset to classify patient's survial chance.
