## Examine Null Values

```python
df = pd.read_csv('../input/beers.csv')
df = df.drop(['Unnamed: 0', 'id'], axis=1)
bf = pd.read_csv('../input/breweries.csv')
df = pd.merge(df, bf, how='left', left_on='brewery_id', right_on='Unnamed: 0')
df = df.drop('Unnamed: 0', axis=1)

plt.figure(figsize=(15,5))
plt.title('Null Values')
ax = sns.heatmap(df.isnull().astype('int'), cmap="inferno")
plt.show()
```

The above figure illustrates that the majority of the ibu column is empty.

```python
plt.figure(figsize=(15,5))
plt.title('Number of Craft Beers per State')
ax = df.state.value_counts().plot()
plt.xticks(np.arange(len(df.state.unique())), df.state.value_counts().index, rotation=80)
plt.show()
```

It appears that Colorado and California have the greatest number of craft beers.

I would like to see what how the ABV amounts are distributed in each state.

```python
plt.figure(figsize=(20,5))
plt.title('Alcohol By Volume By State')
ax = sns.boxplot(x="state", y="abv", data=df)
plt.xticks(np.arange(len(df.state.unique())), df.state.value_counts().index, rotation=80)
plt.show()
```

We see that the interquartile ranges for alcohol by volume are between .04 and .08.

How are the sizes distributed for all beers?

```python
plt.figure(figsize=(10,5))
plt.title('Distribution of Beer Volume')
sns.boxplot(df.ounces)
plt.show()
```

Do larger beers tend to have higher/lower ABV?

```python
plt.figure(figsize=(10,5))
plt.title('ABV vs Volume')
sns.swarmplot(x='ounces', y='abv', data=df)
plt.show()
```

It appears that in general, the answer is no.
