Now we will group the brands and find the num of reviews to each brand and avg rating of that brand. the table will be in ascending order, if you want to know the brands with poor rating change ascending = True.

```python
df.groupby('brand')['reviews.rating'].agg(['mean','count']).sort_values(by='count',ascending=False)[:10]
```
