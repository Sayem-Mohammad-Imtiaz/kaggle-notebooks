```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
url='../input/company-bankruptcy-prediction/'

df=pd.read_csv(url+'data.csv')
df.head()
print(df.columns)
```

```python
df.shape
```

```python
label=df['Bankrupt?']
print(label)
label.value_counts()
```

```python
data=df.drop(['Bankrupt?'],axis=1)
print(data)
```

# Dimension Reduction using PCA

```python
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
Transf_pca = PCA(n_components= 10)
datanew = Transf_pca.fit_transform(data)
print(datanew)
datanew.shape
#datanew shape(6819,10)
```

# Classification Random Forest

```python
label=pd.DataFrame(label)
label.shape
newbase=np.concatenate((datanew,label),axis=1)
print(newbase)
newbase.shape
newbase=pd.DataFrame(newbase)
print(newbase.columns)
```

```python
from sklearn.model_selection import train_test_split
#x_train1,x_test1,y_train1,y_test1=train_test_split(data1,label1,test_size=0.2,random_state=0)
x_train,x_test,y_train,y_test = train_test_split(newbase,label,test_size = 0.2,random_state = 0)

from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
rf = RandomForestClassifier(max_depth=2, random_state=0)
rf.fit(x_train, y_train.values.ravel())
```

```python
x_train.shape
```

# Random Forest F1 score

```python
rfpred=rf.predict(x_test)
from sklearn.metrics import f1_score
f1score=f1_score(y_test,rfpred)*100
print('Random Forest f1_score is: ',f1score) 
```

# Prediction report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,rfpred))
```

# Gradient Boosted Trees

```python
from sklearn.ensemble import GradientBoostingClassifier
gb = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1, random_state=0).fit(x_train, y_train.values.ravel())
gbpred = gb.predict(x_test)
```

# Gradient boosting F1 Score

```python
from sklearn.metrics import f1_score
f1score=f1_score(y_test,gbpred)*100
print('Gradient Boosting f1_score is: ',f1score) 
```

# Report

```python
from sklearn.metrics import classification_report
Class=classification_report(y_test,gbpred)
print(Class)
```

# Extra Trees

```python
from sklearn.ensemble import ExtraTreesClassifier
ext = ExtraTreesClassifier(n_estimators=100, random_state=0)
ext.fit(x_train, y_train.values.ravel())
extpred = ext.predict(x_test)
```

# report

```python
from sklearn.metrics import classification_report
Class=classification_report(y_test,extpred)
print(Class)
```

# Roc curve

```python
from sklearn.metrics import f1_score
f1score=f1_score(y_test,extpred)*100
print('f1_score is: ',f1score) 
```
