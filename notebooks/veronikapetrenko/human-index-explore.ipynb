```python
quant_columns = list()
quant_columns.append(df.columns[0]) #column year
quant_columns.extend(df.columns[4:]) #rest of columns
quant_columns
```

```python
df[quant_columns]
```

```python
rangos = []
for i in quant_columns:
    rango = np.max(df[i]) - np.min(df[i])
    rangos.append(rango)
```

```python
rangos
```

```python
np.max(np.array(df['year'])) - np.min(np.array(df['year']))
```

```python
import scipy
from scipy import stats 
stats.mode(df['ISO_code'])[0][0]
```

```python
for i in qual_columns:
    moda1 = stats.mode(df[i])[0][0]
    print(moda1)
```

```python
medias = []
for i in quant_columns:
    medio = df[i].mean()
    medias.append(medio)
```

```python
medias
```

```python
var = []
st_dev = []
for i in quant_columns:
    var1 = np.var(df[i])
    var.append(var1)
    st = np.std(df[i])
    st_dev.append(st)
```
