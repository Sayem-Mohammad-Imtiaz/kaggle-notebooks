```python
import numpy
import pandas
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier 
from keras.utils import np_utils
from sklearn.model_selection import cross_val_score 
from sklearn.model_selection import KFold
from sklearn.preprocessing import LabelEncoder
from sklearn.pipeline import Pipeline
from keras.utils import to_categorical
```

```python
seed = 7 
numpy.random.seed(seed)
```

```python
# load dataset
dataframe = pandas.read_csv("../input/iris-flower-dataset/IRIS.csv") 
dataframe.head()
```

```python
X=dataframe.drop('species',axis=1)
y=dataframe['species']
dataframe['species'] = pandas.Categorical(dataframe.species)
dataframe['species'] =dataframe.species.cat.codes
# Turn response variable into one-hot response 
y = to_categorical(dataframe.species)
```

```python
dataframe.drop_duplicates(inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,stratify=y,random_state=123)
```

```python
model = Sequential()
model.add(Dense(50, input_dim=4 , activation='relu')) 
model.add(Dense(3, activation='sigmoid'))
# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam' , metrics=['accuracy']) 
```

```python
history=model.fit(X_train,y_train,epochs=100,validation_data=(X_test, y_test))
```
