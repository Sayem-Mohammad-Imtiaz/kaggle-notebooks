```python
# Clustering Analysis
x_mall = coordiate[['Age','Annual Income (k$)','Spending Score (1-100)']]
```

```python
from sklearn.preprocessing import StandardScaler  
scaler = StandardScaler()
scaler.fit(x_mall)
x_mall=scaler.transform(x_mall)
```

```python
def normal_scatter(x,y,z):
    plt.figure(figsize=(8, 4))
    fig = px.scatter_3d(mall, x, y, z,opacity=0.7)
    fig.show()

def cluster_scatter(x,y,z,cluster):
    plt.figure(figsize=(8, 4))
    fig = px.scatter_3d(mall, x, y, z,color=cluster,opacity=0.7)
    fig.show()
```

```python
normal_scatter('Age','Annual Income (k$)','Spending Score (1-100)')
```

```python
from sklearn.cluster import KMeans
```

```python
n_start = 2
n_end = 8

wcss = []
for i in range(n_start, n_end + 1):
    kmeans = KMeans(n_clusters = i, random_state = 0)
    kmeans.fit(x_mall)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(10, 5))
plt.plot(range(n_start, n_end + 1), wcss,marker='o',  
         markerfacecolor='blue', markersize=8)
plt.title('Within cluster sum of squares distance')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') #within cluster sum of squares 
plt.show()
```

```python
# silhouette analysis
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from yellowbrick.cluster import SilhouetteVisualizer
```

```python
def sil_score(x,from_k=2,to_k=6):
    sils=[]
    for k in range(from_k,to_k+1):
        m=KMeans(n_clusters=k)
        m.fit(x)
        silhouette_avg = silhouette_score(x,m.labels_).round(4)
        sils.append([silhouette_avg,k])
    return sils
```

```python
model2=KMeans(n_clusters=2)
visualizer=SilhouetteVisualizer(model2)
visualizer.fit(x_mall)
visualizer.poof()

model3=KMeans(n_clusters=3)
visualizer=SilhouetteVisualizer(model3)
visualizer.fit(x_mall)
visualizer.poof()

model4=KMeans(n_clusters=4)
visualizer=SilhouetteVisualizer(model4)
visualizer.fit(x_mall)
visualizer.poof()

model5=KMeans(n_clusters=5)
visualizer=SilhouetteVisualizer(model5)
visualizer.fit(x_mall)
visualizer.poof()

model6=KMeans(n_clusters=6)
visualizer=SilhouetteVisualizer(model6)
visualizer.fit(x_mall)
visualizer.poof()

model7=KMeans(n_clusters=7)
visualizer=SilhouetteVisualizer(model7)
visualizer.fit(x_mall)
visualizer.poof()

model8=KMeans(n_clusters=8)
visualizer=SilhouetteVisualizer(model8)
visualizer.fit(x_mall)
visualizer.poof()
```
