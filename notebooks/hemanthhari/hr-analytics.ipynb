```python
df_train= pd.read_csv('/kaggle/input/hr-analytics-job-change-of-data-scientists/aug_train.csv')
df_train.head()
```

```python
df_train= df_train.replace([np.inf, -np.inf], np.nan)
df_train= df_train.fillna(0)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
def FuncLabelEncoder(df):
    for c in df.columns:
        if df.dtypes[c] == object:
            le.fit(df[c].astype(str))
            df[c] = le.transform(df[c].astype(str))
    return df
```

```python
train = FuncLabelEncoder(df_train)
train.info()
```

```python
df_test= pd.read_csv('../input/hr-analytics-job-change-of-data-scientists/aug_test.csv')
df_test.head()
```

```python
df_test = FuncLabelEncoder(df_test)
df_test.info()
```

```python
features =["enrollee_id","city","city_development_index","gender","relevent_experience","enrolled_university","education_level","major_discipline","experience","company_size","company_type","last_new_job","training_hours"]
target = 'target'
```

```python
df_train[features].head(10)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import r2_score, mean_squared_error

# We define the model
clf = XGBClassifier()
# We train model
clf.fit(df_train[features],df_train[target])
```

```python
pred = clf.predict(df_test[features])
pred
```

```python
submission = pd.DataFrame({'enrollee_id':df_test['enrollee_id'],'target':pred})
submission.head(20)
```

```python
filename = 'submission.csv'
submission.to_csv(filename,index=False)
print('Saved file: ' + filename)
```
