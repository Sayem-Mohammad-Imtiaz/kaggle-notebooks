***Count the empty rows in each columns***

```python
for column in data.columns:
    print("No of empty rows in {} is {}".format(column,data[column].isna().sum()))
```

***Extracting columns with null values***

```python
empty_features = [column for column in data.columns if data[column].isna().sum() > 0]
empty_features
```

***Calculate the percentage of null values in above columns***

```python
for feature in empty_features:
    df = data.copy()
    print('Percentage of empty rows in {} is {} %'.format(feature,np.round(df[feature].isna().mean(),2)))
```

```python
df.info()
```

***Checking the unique values & count of missing values in Basepay***

```python
df["BasePay"].unique()
```

```python
df["BasePay"].isna().sum()
```

***Replacing the null values & "Not Provided" values in Basepay & Benefits with 0s***

```python
for column in ['BasePay', 'Benefits']:
    df[column] = np.where((df[column].isna()) | (df[column] == "Not Provided"),
                          0,
                          df[column])
```

```python
df["BasePay"].isna().sum()
```

***Dropping "Id","Notes" & "Agency" columns***

```python
df.drop(['Id','Agency','Notes'],axis = 1, inplace = True)
```

***Replacing the NaN values with new category "Missing Values"***

```python
df["Status"] = df["Status"].fillna("Missing Value")
```
