```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.ticker as mtick  
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import os

hotel_data = pd.read_csv('../input/hotel-booking-demand/hotel_bookings.csv')
```

```python
hotel_data.head(5)
```

```python
data = hotel_data[hotel_data['is_canceled'] == 0]
fig = plt.figure(figsize=(9,6), dpi = 100)
boxplot = sns.boxplot(data= hotel_data, x = 'reserved_room_type', y = 'adr', hue = 'hotel')
boxplot.set_ylim(0, 500)
```

Check for the percentage of missing values in the dataset

```python
missing = pd.DataFrame((hotel_data.isnull().sum())*100/hotel_data.shape[0]).reset_index()
plt.figure(figsize=(16,5))
ax = sns.pointplot('index',0,data=missing)
plt.xticks(rotation =90,fontsize =7)
plt.title("Percentage of Missing values")
plt.ylabel("PERCENTAGE")
plt.show()
```

```python
hotel_count = sns.countplot(data=hotel_data, x = 'hotel', hue = 'hotel')

hotel_count.set_title('No. of Resort Hotel and City Hotel')

hotel_data['hotel'].value_counts()
```

```python
is_canceled_count = sns.countplot(data=hotel_data, x = 'is_canceled', hue = 'is_canceled')

is_canceled_count.set_title('No. of canceled')

hotel_data['is_canceled'].value_counts()
```

```python
is_canceled_hotels = sns.countplot(data=hotel_data, x = 'hotel', hue='is_canceled')

hotel_data.groupby('hotel')['is_canceled'].value_counts()
```

```python
Mth = sns.kdeplot(hotel_data.lead_time[(hotel_data["is_canceled"] == 0) ],
                color="Green", shade = True)
Mth = sns.kdeplot(hotel_data.lead_time[(hotel_data["is_canceled"] == 1) ],
                ax =Mth, color="Red", shade= True)
Mth.legend(["No Cancel","Cancel"],loc='upper right')
Mth.set_ylabel('Density')
Mth.set_xlabel('Lead Time')
Mth.set_title('Lead Time by Is_canceled')
```
