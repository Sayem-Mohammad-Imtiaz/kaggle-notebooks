# Ordinal Encoder

By using Ordinal Encoder, we can change the data type of an object to numerical.
Because of all the dataset is an object, both X and y is transformed by Ordinal Encoder.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import mean_absolute_error
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder

X = dataset.drop(['class'],axis=1)
y = dataset['class']

categorical_cols_X = [cname for cname in X.columns if 
                    X[cname].dtype == "object"]
good_label_cols = ['class']

label_y_train = dataset.drop(categorical_cols_X, axis=1)
ordinal_encoder = OrdinalEncoder()
label_y_train[good_label_cols] = ordinal_encoder.fit_transform(dataset[good_label_cols])

X_train, X_test, y_train, y_test = train_test_split(X,label_y_train,train_size=0.8, test_size=0.2,random_state=0)

label_X_train = X_train
label_X_test = X_test
label_X_train[categorical_cols_X] = ordinal_encoder.fit_transform(X_train[categorical_cols_X])
label_X_test[categorical_cols_X] = ordinal_encoder.transform(X_test[categorical_cols_X])
```

```python
X_test
```

```python
y_train.tail()
```

```python
model = RandomForestClassifier()
model.fit(label_X_train, y_train)
preds = model.predict(label_X_test)
print(mean_absolute_error(y_test, preds))
```
