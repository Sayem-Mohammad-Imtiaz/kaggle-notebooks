```python
#setup
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
#setup part 2
netflix_filepath = "../input/netflix-shows/netflix_titles.csv"

netflix_data = pd.read_csv(netflix_filepath, index_col="show_id", parse_dates=True)
print("Setup Complete")
```

This is the setup to get kaggle to recognise the netflix title data set

```python
netflix_country = netflix_data["country"].value_counts().head(15)
#netflix_country = netflix_country.head(10)
```

This makes a variable that counts all the countries and shows the value

```python
#test, Here is the data of shows produced in the top 15 countries 
#netflix_data.tail(10)
netflix_country
```

```python
# Graph showing top 15 countries that produces the most shows 
# Set the width and height of the figure
plt.figure(figsize=(20,10))

# Add title
plt.title("Amount of shows produced by country")

# Bar chart showing how many shows are produced per country on netflix
sns.barplot(x=netflix_country.index, y=netflix_country)

# Add label for vertical and horizontal axis
plt.ylabel("Amount of shows produced")
plt.xlabel("Countries")

# Style of graph
sns.set_style("darkgrid")
```

This graph clearly shows that most shows on Netflix are produced by the United states of America. I just chose the top 15 countries as the graph already shows where most shows are produced and I don't think there would really be a point in showing the rest of the countries. 

```python
netflix_releaseyear = netflix_data["release_year"].value_counts().head(15)
```

```python
#test
netflix_releaseyear
```

```python

# Graph
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Release year")

# Line graph showing amount of shows being released, yearly
sns.lineplot(x=netflix_releaseyear.index, y=netflix_releaseyear)

# Add label for vertical and horizontal axis
plt.ylabel("Shows")
plt.xlabel("Year")

# Style of graph 
sns.set_style("darkgrid")
```

The streaming service started in 2007 so there isn't a point going before that.
This graph shows the increasing popularity of Netflix which allowed it to release more movies and shows
