```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

> ***How Many More Movies Are Released In The Past 20 Years And Is It More Then How Many TV Shows Are Released?***

```python
netflix_filepath = "../input/kaggle-data/Data Kaggle.csv"

netflix_data = pd.read_csv(netflix_filepath, index_col="type")


netflix_Movies = netflix_data.loc['Movie']
netflix_TV = netflix_data.loc['TV Show']

netflix_release_year = netflix_Movies["release_year"].value_counts().head(100)
netflix_release_year2 = netflix_TV["release_year"].value_counts().head(20)
```

```python
#netflix_data.head()
#netflix_release_year
netflix_TV
```

```python
plt.figure(figsize=(20,10))

plt.title("Movies")

plt.xlabel("Year", fontsize = 50)
plt.title("Movies", fontsize = 50)
plt.ylabel("Amount Of Movies", fontsize = 50)

sns.barplot(x=netflix_release_year.index, y=netflix_release_year)

plt.ylabel("Amount Of Movies")
plt.xlabel("Years")
```

```python
plt.figure(figsize=(20,10))

plt.title("TV Shows")

plt.xlabel("Year", fontsize = 50)
plt.title("TV Shows", fontsize = 50)
plt.ylabel("Amount Of TV Shows", fontsize = 50)

sns.barplot(x=netflix_release_year2.index, y=netflix_release_year2)

plt.ylabel("Amount Of TV Shows")
plt.xlabel("Years")
```

I wanted to find out how many movies and TV shows were created in the span of 20 years starting from 2000 - 2020. With these 2 graphs I can clearly see the year when most movies were made and the exact same thing but with TV Shows.
