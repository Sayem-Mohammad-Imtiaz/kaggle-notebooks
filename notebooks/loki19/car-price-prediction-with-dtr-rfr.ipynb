```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import RFE
from sklearn.metrics import r2_score

sns.set()
```

```python
data = pd.read_csv('../input/car-price-prediction/CarPrice_Assignment.csv')
data.head()
```

**Dropping the car ID**

# drop car_ID

```python
data.drop('car_ID',axis=1,inplace=True)
```

```python
data.head()
```

```python
data.info()
```

```python
data['CarName'] = data['CarName'].str.split(' ',expand=True)
data.head()
```

car name duplication like(nissan and Nissan are same, toyota and toyouta are same, vokswagen and volkswagen and vw are same, mazda and maxda are same, porcshce and porsche are same


```python
data['CarName'] = data['CarName'].replace({'maxda': 'mazda', 'nissan': 'Nissan', 'porcshce': 'porsche', 'toyouta': 'toyota', 
                            'vokswagen': 'volkswagen', 'vw': 'volkswagen'})
```

```python
data['symboling']=data['symboling'].astype('str')
```

```python
X = data.apply(lambda col: preprocessing.LabelEncoder().fit_transform(col))
X=X.drop(['CarName','price'],axis=1)
y=data['price']
```

```python
X.head()
```

# Decision tree

```python
x_train,x_test,y_train,y_test = train_test_split(X,y, random_state = 0)
```

```python
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(x_train,y_train)
y_train_pred = regressor.predict(x_train)
y_test_pred = regressor.predict(x_test)
regressor.score(x_test,y_test)
```

```python
train = pd.DataFrame({'y_train':y_train,'y_train_pred':y_train_pred})
train
```

```python
Rand_frst = RandomForestRegressor(criterion = 'mse',
                              random_state = 20,
                              n_jobs = -1)
Rand_frst.fit(x_train,y_train)
Rand_frst_train_pred = Rand_frst.predict(x_train)
Rand_frst_test_pred = Rand_frst.predict(x_test)


r2_score(y_test,Rand_frst_test_pred)
```

# Conclusion

We applied two models Decision Tree Regressor,and RandomForest Regressor because half of the independent columns are categorical
