```python
# CONVERT SALARY FROM TEXT TO NUMBERS WITH DUMMIES
```

```python
x_salary_dummies = pd.get_dummies(x1['salary'])
```

```python
x = x_with_dummies.drop('salary', axis=1)
y = raw_data['left']
x
```
