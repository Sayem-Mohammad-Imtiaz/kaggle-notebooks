# Features and Target

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import r2_score,classification_report, confusion_matrix,accuracy_score,f1_score
```

```python
X = diabetes.drop('Outcome',axis=1)
y = diabetes['Outcome']

scaler = StandardScaler()
scaler.fit(X)
x = scaler.transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=101)
```

```python
# LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
predict_lr = lr.predict(X_test)

print('r2_score:',r2_score(y_test,predict_lr))
```

```python
# Logistic Regression
```

```python
log_r = LogisticRegression()
log_r.fit(X_train,y_train)
predict_logr = log_r.predict(X_test)

print('Confusion matrix')
print(confusion_matrix(y_test,predict_logr))
print('classification_report')
print(classification_report(y_test,predict_logr))
print('accuracy_score:',accuracy_score(y_test,predict_logr))
print('f1_score:',f1_score(y_test,predict_logr))
```

```python
# Decision Tree
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
predict_dt = dt.predict(X_test)

print('Confusion matrix')
print(confusion_matrix(y_test,predict_dt))
print('classification_report')
print(classification_report(y_test,predict_dt))
print('accuracy_score:',accuracy_score(y_test,predict_dt))
print('f1_score:',f1_score(y_test,predict_dt))
```

```python
# Random Forest
```

```python
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train,y_train)
predict_rfc = rfc.predict(X_test)

print('Confusion matrix')
print(confusion_matrix(y_test,predict_rfc))
print('classification_report')
print(classification_report(y_test,predict_rfc))
print('accuracy_score:',accuracy_score(y_test,predict_rfc))
print('f1_score:',f1_score(y_test,predict_rfc))
```

```python
# K Nearest Neighbors
```

```python
error_rate = []
for i in range(1,40):
    k = KNeighborsClassifier(n_neighbors=i)
    k.fit(X_train,y_train)
    predict_i = k.predict(X_test)
    error_rate.append(np.mean(predict_i != y_test))
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(1,40),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
```

```python
knn = KNeighborsClassifier(n_neighbors=9)
knn.fit(X_train,y_train)
predict_knn = knn.predict(X_test)

print('Confusion matrix')
print(confusion_matrix(y_test,predict_knn))
print('classification_report')
print(classification_report(y_test,predict_knn))
print('accuracy_score:',accuracy_score(y_test,predict_knn))
print('f1_score:',f1_score(y_test,predict_knn))
```

```python
# Support Vector Machine
```

```python
svm = SVC()
svm.fit(X_train,y_train)
predict_sv = svm.predict(X_test)

print('Confusion matrix')
print(confusion_matrix(y_test,predict_sv))
print('classification_report')
print(classification_report(y_test,predict_sv))
print('accuracy_score:',accuracy_score(y_test,predict_sv))
print('f1_score:',f1_score(y_test,predict_sv))
```

```python
# Grid Search
```

```python
param_svm = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001],'kernel':['linear','poly','rbf','sigmoid']}
```

```python
grid_svm = GridSearchCV(SVC(),param_svm,verbose=3)
grid_svm.fit(X_train,y_train)
```

```python
grid_svm.best_params_
```

```python
grid_svm.best_score_
```
