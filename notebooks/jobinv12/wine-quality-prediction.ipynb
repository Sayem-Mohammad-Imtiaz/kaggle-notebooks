```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report, r2_score
```

```python
X = wine.drop('quality',axis=1)
y = wine['quality']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Linear Regression

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
predict_lr = lr.predict(X_test)

print(r2_score(y_test,predict_lr))
```

# Logistic Regression

```python
log_r = LogisticRegression()
log_r.fit(X_train,y_train)
predict_logr = log_r.predict(X_test)

print(confusion_matrix(y_test,predict_logr))
print("\n")
print(classification_report(y_test,predict_logr))
```

# Decision Tree

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
predict_dtc = dtc.predict(X_test)

print(confusion_matrix(y_test,predict_dtc))
print("\n")
print(classification_report(y_test,predict_dtc))
```

# Random Forest

```python
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_train,y_train)
predict_rfc = rfc.predict(X_test)

print(confusion_matrix(y_test,predict_rfc))
print("\n")
print(classification_report(y_test,predict_rfc))
```

# K Nearest Neighbors

```python
kn = KNeighborsClassifier(n_neighbors =1)
kn.fit(X_train,y_train)
predict_k = kn.predict(X_test)

print(confusion_matrix(y_test,predict_k))
print("\n")
print(classification_report(y_test,predict_k))
```

```python
error_rate = []
for i in range(1,40):
    k = KNeighborsClassifier(n_neighbors = i)
    k.fit(X_train,y_train)
    predict_i = k.predict(X_test)
    error_rate.append(np.mean(predict_i != y_test))
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(1,40),error_rate,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize=10)
```

# Support Vector Machine

```python
svm = SVC()
svm.fit(X_train,y_train)
predict_svm = svm.predict(X_test)

print(confusion_matrix(y_test,predict_svm))
print("\n")
print(classification_report(y_test,predict_svm))
```
