```python
correlation = f.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, vmax=1, square=True,annot=True,cmap='cubehelix')

plt.title('Analysis of Game of Throne')

```

Conclusion from the heatmap: there is a strong positive correlation between death and the time of the year, which is summer (correlation is positive 0.34). There is strong negative correlation between summer and number of battles fought (we may assume that most battles were fought in winter). 

```python
f.hist()
```

Conclusion: there is a strong positive correlation between death and the time of the year, which is summer (correlation is positive 0.34). More deaths happen in summer time. There seems to be negative correlation between the number of battle fought and the time of summer. 

Now let's see which attacker won the most battles.

```python
f.groupby(['attacker_king', 'attacker_outcome']).count()['name'].unstack().plot(kind='bar')
```

The above graph shows that attacker Joffrey/Tommen Baratheon won the most battles.  Attacher Balon/Euron Greyjoy only had wins, they never lost a battle.
