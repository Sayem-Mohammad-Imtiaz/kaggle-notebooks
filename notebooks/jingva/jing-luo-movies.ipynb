Which director makes the most movies?
Is there a correlation between facebook likes and gross?
Is there a correlation between facebook likes and imdb ratings?

```python
f=pd.read_csv("../input/movie_metadata.csv")
```

```python
f.head()
```

```python
f.shape
```

```python
f.corr()
```

```python
correlation = f.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, vmax=1, square=True,annot=True,cmap='cubehelix')

plt.title('Movies Analysis')
```

From the heatmap, we can see that there is strong positive correlation between voted users and gross (0.64), user review and gross(0.56), critics review and gross (0.48).  There is less correlation between actor facebook likes and imdb ratings.

Let's run another test of the data.  Let's see which director has the most facebook likes.

```python
chart = f.groupby(["director_name"])['director_facebook_likes'].sum()
chart.plot(figsize=(10,5))
chart.plot()
```

```python
The above chart shows that Mulcahy has the most facebook likes.
```

Now I'm making the data set smaller for further analysis.

```python
msample = f[['director_name','gross','imdb_score']]
msample.head()
```

```python
msample_group = msample.groupby('director_name')
msample_group.size()
```

Here we can see the director Woody Allen has made the most movies as he has the most counts of 22. The second most movies are Adam Shankman (8 movies), Yimou Zhang (8 movies) and Zack Snyder (8 movies).  

```python
gross_total = msample_group.sum()
gross_total.head()
```

```python
movie_chart = gross_total.plot()
movie_chart.plot(figsize=(50,30))
movie_chart.plot()
```

The above chart shows that Mulcahy makes the most gross.
