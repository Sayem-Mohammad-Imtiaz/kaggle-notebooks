```python
df = pd.read_csv('/kaggle/input/titanicdataset-traincsv/train.csv')
avg_age = df["Age"].astype("float").mean(axis=0)
df["Age"].replace(np.nan, avg_age, inplace = True)
df["Age"] = df["Age"].astype("float")
df.info()
df.describe()
df.head()
```

```python
df=df.drop(['Name', 'Ticket', 'Cabin', 'PassengerId'], axis=1)
df = pd.get_dummies(df,columns=['Sex'],drop_first=True)
df = pd.get_dummies(df,columns=['Embarked'])
df.info()
```

```python
sns.heatmap(df.corr(),cmap='coolwarm')
plt.title('titanic.corr()')
```

```python
X=df.drop(['Survived'],axis=1)
y=df['Survived']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
clf1 = SVC(kernel='linear').fit(X_train, y_train)

print(classification_report(y_test,clf1.predict(X_test)))
print('conf matrix:\n',confusion_matrix(y_test,clf1.predict(X_test)))
```

```python
clf2 = SVC(kernel='sigmoid').fit(X_train, y_train)

print(classification_report(y_test,clf2.predict(X_test)))
print('conf matrix:\n',confusion_matrix(y_test,clf2.predict(X_test)))
```

```python
clf2 = SVC(kernel='sigmoid',coef0=-2.25).fit(X_train, y_train)

print(classification_report(y_test,clf2.predict(X_test)))
print('conf matrix:\n',confusion_matrix(y_test,clf2.predict(X_test)))
```
