let us compare scores of boys and girls between math, reading and writing

```python
fig, ax = plt.subplots(3, sharex=True, )
ax[0].bar('female', df[df['gender'] == 'female']['math score'].mean(), )
ax[0].bar('male', df[df['gender'] == 'male']['math score'].mean(), )
ax[1].bar('female', df[df['gender'] == 'female']['reading score'].mean(), )
ax[1].bar('male', df[df['gender'] == 'male']['reading score'].mean())
ax[2].bar('female', df[df['gender'] == 'female']['writing score'].mean())
ax[2].bar('male', df[df['gender'] == 'male']['writing score'].mean())
ax[2].set_xlabel('gender')
ax[0].set_ylabel('Math mean marks', rotation=90)
ax[1].set_ylabel('Reading mean marks')
ax[2].set_ylabel('Writing mean marks')
plt.xticks(rotation=90)
plt.show()
```

From above we see that in Mathematics the mean marks of boys is more, but in case of reading and writing the mean score of girls is much better than boys.

Let us also see the maximum and minimum marks for the three subjects in case of boys and girls

```python
fig, ax = plt.subplots(3, 2)
ax[0, 0].bar('female', df[df['gender'] == 'female']['math score'].max(), )
ax[0, 0].bar('male', df[df['gender'] == 'male']['math score'].max(), )
ax[0, 1].bar('female', df[df['gender'] == 'female']['math score'].min(), )
ax[0, 1].bar('male', df[df['gender'] == 'male']['math score'].min(), )
ax[1, 0].bar('female', df[df['gender'] == 'female']['reading score'].max(), )
ax[1, 0].bar('male', df[df['gender'] == 'male']['reading score'].max())
ax[1, 1].bar('female', df[df['gender'] == 'female']['reading score'].min(), )
ax[1, 1].bar('male', df[df['gender'] == 'male']['reading score'].min())
ax[2, 0].bar('female', df[df['gender'] == 'female']['writing score'].max())
ax[2, 0].bar('male', df[df['gender'] == 'male']['writing score'].max())
ax[2, 1].bar('female', df[df['gender'] == 'female']['writing score'].min())
ax[2, 1].bar('male', df[df['gender'] == 'male']['writing score'].min())
plt.show()
```

From this we see that the maximum score of boys in all three subjects is at the same level with that to girls, but the minimum marks in each subject is greater than the minimum marks of girls in each subject

Now let us see if some student have completed the test preparation course, is he she better than the students who haven't completed the preperation course

Let us add a new column to total all marks to also see the overall effect on all subject from the preparation course

```python
df['sum of all scores'] = df['math score'] + df['reading score'] + df['writing score']
```

```python
fig, ax = plt.subplots(4, sharex=True)
ax[0].bar('Completed', df[df['test preparation course'] == 'completed']['sum of all scores'].mean())
ax[0].bar('Not Completed', df[df['test preparation course'] != 'completed']['sum of all scores'].mean())
ax[0].set_ylabel('Overall')
ax[1].bar('Completed', df[df['test preparation course'] == 'completed']['math score'].mean())
ax[1].bar('Not Completed', df[df['test preparation course'] != 'completed']['math score'].mean())
ax[1].set_ylabel('Math')
ax[2].bar('Completed', df[df['test preparation course'] == 'completed']['reading score'].mean())
ax[2].bar('Not Completed', df[df['test preparation course'] != 'completed']['reading score'].mean())
ax[2].set_ylabel('Reading')
ax[3].bar('Completed', df[df['test preparation course'] == 'completed']['writing score'].mean())
ax[3].bar('Not Completed', df[df['test preparation course'] != 'completed']['writing score'].mean())
ax[3].set_ylabel('Writing')
```

From the above plot we saw that the students who have completed their test preperation course outperforms other students in all subjects and also overall. **So, I think this preperation course should be made mandatory for all courses**
