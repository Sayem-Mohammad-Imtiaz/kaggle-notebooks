```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report
from catboost import CatBoostClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
```

 # Correlation matrix

```python
plt.figure(figsize=(15,6))
ax = sns.heatmap(df.corr(),cmap='coolwarm')
```

# Plot the high correlation alcohol vs quality

```python
fig, (ax1, ax2) = plt.subplots(1, 2,
                              figsize=(15,6))
ax1.hist(df.alcohol[(df.quality < 6)])
ax1.set_title("Alcohol amount in bad quality wine")
ax1.set_xlabel('Alcohol amount', fontsize='x-large')
ax1.set_ylabel('Number of samples', fontsize='x-large')
ax2.hist(df.alcohol[(df.quality > 6)])
ax2.set_title('Alcohol amount in good quality wine')
ax2.set_xlabel('Alcohol amount', fontsize='x-large')
ax2.set_ylabel('Number of samples', fontsize='x-large');
```

# Plot the value count for each quality measure

```python
df.quality.value_counts().plot(kind='bar');
```

**We can see that quality 5 and 6 are the most dominant.**

# Converting the quality column into binary
* 0 for bad wine
* 1 for good wine

```python
df.quality = [1 if  i > 6 else 0 for i in df.quality]
```

# Preparing the data ( train and test split)

```python
X_wine = df.drop('quality', axis=1)
y_wine = df['quality']
X_train, X_test, y_train, y_test = train_test_split(X_wine, y_wine, test_size=0.2, random_state=42)
```

# Scaling the data with StandardScaler

```python
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

# Fit and score 

```python
cat = CatBoostClassifier(silent=True)
cat.fit(X_train, y_train)
y_preds = cat.predict(X_test)
print(classification_report(y_test, y_preds, zero_division=0))
```
