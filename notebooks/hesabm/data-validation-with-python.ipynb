* ENG: When the values in the data set are examined, zero values are seen in the Glucose, Blood Pressure, Skin Thickness, Insulin and BMI columns. This erroneous data must be deleted. Because these data are not true data.

* TR: Veri setindeki değerler incelendiğinde Glucose, Blood Pressure, Skin Thickness, Insulin ve BMI sütunlarında sıfır değerleri görülmektedir. Bu hatalı verilerin silinmesi gerekmektedir. Çünkü bu veriler doğru veriler değildir.

```python
filtered_data = data[data["Glucose"] == 0 ] 
filtered_data.index.values
```

```python
filtered_data1 = data[data["BloodPressure"] == 0 ] 
filtered_data1.index.values
```

```python
filtered_data2 = data[data["SkinThickness"] == 0 ] 
filtered_data2.index.values
```

```python
filtered_data3 = data[data["BMI"] == 0 ] 
filtered_data3.index.values
```

```python
filtered_data4 = data[data["Insulin"] == 0 ] 
filtered_data4.index.values
```

ENG: The index of all values that need to be deleted have been added to a list.

TR: Silinmesi gereken tüm değerlerin indexleri bir listeye eklenmiştir.

```python
all_list = filtered_data + filtered_data1 + filtered_data2 + filtered_data3 + filtered_data4

# Number of incorrect data
# Hatalı veri sayısı
incorrect = len(all_list)

# Total number of data
# Toplam veri sayısı
total = len(data)

# Correct number of data
# Doğru veri sayısı
correct = total - incorrect



# Pie chart, where the slices will be ordered and plotted counter-clockwise:
# Dilimlerin sıralanacağı ve saat yönünün tersine çizileceği pasta grafik

labels = "Correct Data", "Incorrect Data"
sizes = (incorrect, correct)
explode = (0, 0.1) 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')

plt.show()
```
