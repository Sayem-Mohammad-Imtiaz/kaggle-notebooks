## State wise Indian Covid-19 Dataset Data Exploration

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('/kaggle/input/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv')
```

### Basic Stats Regarding This dataset:

```python
print(f"Average Active Ratio is {df['Active Ratio (%)'].mean()}%")
print(f"Average Discharge Ratio is {df['Discharge Ratio (%)'].mean()}%")
print(f"Average Death Ratio is {df['Death Ratio (%)'].mean()}%")
```

This Data shows that the Discharge means, Recover Ratio is About 96.05%, with Respect to Death Ratio is 1.26%

```python
df.describe()
```

```python
df = df.sort_values(by='Death Ratio (%)',ascending=False)
plt.figure(figsize=(13,15))
sns.barplot(y = 'State/UTs',x = 'Death Ratio (%)',data=df,color='Black')
plt.ylabel('State')
plt.xlabel('Death Ratio (%)')
plt.title('Death Ratio in Different States')
```

```python
df = df.sort_values(by='Discharge Ratio (%)',ascending=False)
plt.figure(figsize=(13,15))
sns.barplot(y = 'State/UTs',x = 'Discharge Ratio (%)',data=df,color='Black')
plt.ylabel('State')
plt.xlabel('Discharge Ratio (%)')
plt.title('Discharge Ratio in Different States')
```

```python
df = df.sort_values(by='Active Ratio (%)',ascending=False)
plt.figure(figsize=(13,15))
sns.barplot(y = 'State/UTs',x = 'Active Ratio (%)',data=df,color='Black')
plt.ylabel('State')
plt.xlabel('Active Ratio (%)')
plt.title('Active Ratio in Different States')
```

we have seen in Death Ratio and Active is not Uniformly Distributed, but the Discharge Ratio or Recovery Ratio is approximatly Uniformly Distributed Amoung All State, But the Best State is the Covid Time is Dadra and Nagar Haveli and Daman and Diu, which has the Highest Discharge Rates as well as having lowest Death Rates, Make this Very Successfull amoung All State is Recovering and Dealing with Covid 19,

But the Mizoram has the Lowest Discharge Ratio and Highest Active Ratio.
