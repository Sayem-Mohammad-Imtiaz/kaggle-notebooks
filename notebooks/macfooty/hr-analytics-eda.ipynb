Graduates are more likely to start looking for better jobs compared to data scientists who have masters degree.

```python
data.groupby('target')['enrolled_university'].value_counts(normalize=True).unstack().plot.bar(stacked=True)
```

If a data scientist is enrolled in a full-time program, he's more likely to start looking for a new job. Whereas if he is not enrolled in any university program, he's more likely to stay.
