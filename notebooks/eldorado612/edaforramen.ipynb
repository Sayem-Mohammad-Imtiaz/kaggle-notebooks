```python
df = pd.read_csv("../input/ramen-ratings/ramen-ratings.csv")
df.head()
```

```python
def resumetable(df):
    print(f"Dataset Shape: {df.shape}")
    summary = pd.DataFrame(df.dtypes,columns=['dtypes'])
    summary = summary.reset_index()
    summary['Name'] = summary['index']
    summary = summary[['Name','dtypes']]
    summary['Missing'] = df.isnull().sum().values    
    summary['Uniques'] = df.nunique().values
    summary['First Value'] = df.loc[0].values
    summary['ValueCount']= df.count().values
    return summary
```

```python
resumetable(df)
```

```python
#Replacing the NA value by the most common style by respective country
df.loc[(df["Style"].isnull()) & (df["Country"] == "China"),"Style"] = df[df["Country"] == "China"].Style.value_counts().keys()[0]
df.loc[(df["Style"].isnull()) & (df["Country"] == "Taiwan"),"Style"] = df[df["Country"] == "Taiwan"].Style.value_counts().keys()[0]
```

```python
def makeinteger(e):
    try:
        return float(e)
    except:
        return None
```

```python
df.Stars = df.Stars.apply(makeinteger)
```

```python
df["Stars"] = df["Stars"].fillna(df["Stars"].mean())
```
