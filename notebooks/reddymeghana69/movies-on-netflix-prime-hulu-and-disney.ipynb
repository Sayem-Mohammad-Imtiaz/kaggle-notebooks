```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("../input/movies-on-netflix-prime-video-hulu-and-disney/MoviesOnStreamingPlatforms_updated.csv")
```

```python
df.shape
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
plt.subplot(221)

df['Prime Video'].value_counts().plot(kind='bar', title='Amazon Prime Video', figsize=(16,9))

plt.xticks(rotation=0)

plt.subplot(222)

df['Disney+'].value_counts().plot(kind='bar', title='Disney+')

plt.xticks(rotation=0)

plt.subplot(223)

df['Hulu'].value_counts().plot(kind='bar', title='Hulu')

plt.xticks(rotation=0)

plt.subplot(224)

df['Netflix'].value_counts().plot(kind='bar', title='Netflix')

plt.xticks(rotation=0)

plt.show()
```

```python
df.boxplot()
```

```python
sns.distplot(df['Runtime'])
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True, square=True)
plt.yticks(rotation=2)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.scatter(df["Hulu"], df["Netflix"])
ax.set_xlabel("Hulu")
ax.set_ylabel("Netflix")
plt.show()
```
