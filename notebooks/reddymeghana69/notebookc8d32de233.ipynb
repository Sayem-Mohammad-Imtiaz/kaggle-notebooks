```python
plt.subplot(221)

df['inning'].value_counts().plot(kind='bar', title='inning', figsize=(16,9))

plt.xticks(rotation=0)

plt.subplot(222)

df['over'].value_counts().plot(kind='bar', title='over')

plt.xticks(rotation=0)

plt.subplot(223)

df['ball'].value_counts().plot(kind='bar', title='ball')

plt.xticks(rotation=0)

plt.subplot(224)

df['batsman'].value_counts().plot(kind='bar', title='batsman')

plt.xticks(rotation=0)

plt.show()
```
