```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv")
```

```python
plt.subplot(221)

df['Reviews'].value_counts().plot(kind='bar', title='Reviews', figsize=(16,9) , color="green")

plt.xticks(rotation=0)

plt.subplot(222)

df['Price'].value_counts().plot(kind='bar', title='Price', color="pink")

plt.xticks(rotation=0)

plt.subplot(223)

df['Year'].value_counts().plot(kind='bar', title='Year', color="violet" )

plt.xticks(rotation=0)

plt.subplot(224)

df['Genre'].value_counts().plot(kind='bar', title='Genre',color="orange")

plt.xticks(rotation=0)

plt.show()
```

```python
df.boxplot()
```

```python
sns.heatmap(df.isnull())
```

```python
sns.distplot(df['Price'])
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True, square=True)
plt.yticks(rotation=2)
plt.show()
```

```python
df.hist(figsize=(10, 12), bins=50, xlabelsize=8, ylabelsize=8 ); 
```

```python
sns.relplot(x="User Rating", y="Price", hue='Genre', data=df)
```

```python
sns.factorplot("Reviews",data=df,kind="count")
```

```python
sns.factorplot("Price",data=df,kind="count")
```

```python
sns.set_palette("gist_rainbow_r")
sns.jointplot(x="Price",y="User Rating",kind="hex",data=df)
plt.show
```
