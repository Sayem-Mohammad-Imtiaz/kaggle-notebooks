# **Pre-process the data**

```python
# Drop unimportant covariates
#here we drop all the variables that are not helpful in the model building
df.drop(columns=['interviewer', 'vname','Unnamed: 0' ],axis=1,inplace=True)

# Step 2: Drop ALL redundant covariates

df.drop(columns=['adjtempshifts1_1','adjtempshifts2_1', 'adjtempshifts3_1','adjrainfallshifts1_1','adjrainfallshifts2_1','adjrainfallshifts3_1'], axis=1, inplace=True)
```

```python
# drop longtermrainfallshifts3 since it does not contain any record for the whole dataset.
df.drop(columns=['longtermrainfallshifts3' ],axis=1,inplace=True)
```

```python
#fill the non values.
df = df.fillna(0)
df.head()
```

```python
low = 0.05
high = 0.95

# Step 1: compute 5% percentile and the 95% percentile of each column in the dataset
quantile_df = df.quantile([low, high])
quantile_df
```

# Visualizations

```python
#pie chart visualization for gender1
ftype = df['gender1']
List = [1,2]
Count = [sum(map(lambda x: x==i,ftype)) for i in List]
Perc = [x*100.0/sum(Count) for x in Count]
plt.pie(Perc,None,List,autopct='%1.1f%%')
plt.show()
```
