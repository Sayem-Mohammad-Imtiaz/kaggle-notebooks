```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.neural_network import MLPClassifier

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv('../input/mushroom-classification/mushrooms.csv')
data.describe()
```

veil-type має лише одне унікальне значення, тому його можна видалити з датасету

```python
data = data.drop('veil-type', axis = 1)
```

для роботи з даним датасетом всі дані мають бути в чисельному вигляді, тому треба використати кодування

```python
classes = data.columns
for i in classes:
    data[i] = LabelEncoder().fit_transform(data[i])
data.describe()
```

```python
data.head()
```

тепер необхідно дізнатися "найважливіші" змінні, для отримання кореляції між всіма парами змінних використовуємо heatmap:

```python
plt.figure(figsize = (15,12), dpi = 90)
sns.heatmap(data.corr(), cmap = 'RdYlGn', center = 0, annot = True)
plt.show()
```

bruises, gill-size та gill-color найбільше корелюють із значенням класу грибів

```python
# розділення даних на тренувальні та тестувальні у співвідношенні 1:1

X = data.drop('class', axis = 1)
Y = data['class']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.5)
```

спробуємо декілька функцій активації та оберемо найкращий результат

```python
sigm = MLPClassifier(max_iter = 250, activation = 'logistic').fit(X_train, Y_train) # сигмоїда
```

```python
print(classification_report(sigm.predict(X_test), Y_test))
```

```python
relu = MLPClassifier(max_iter = 250, activation = 'relu').fit(X_train, Y_train) # relu
```

```python
print(classification_report(relu.predict(X_test), Y_test))
```

як бачимо, модель добре спрацювала з різними функціями активації, але якщо дані не були б досить точними, можливо ми б отримали суттєву різницю між різними функціями
