```python
location_data=space_df.groupby('location',as_index=False)['datum'].count().sort_values(by='datum',ascending=False)
location_data.columns=['location','count']
location_data.reset_index(drop=True,inplace=True)
l1=list(location_data['location'][0:200])
l2=list(location_data['count'][0:200])
```

<h2><b>Lets plot the top 50 locations used for launching satellites</b></h2>

```python
import folium
```

```python
world_map=folium.Map(zoom_start=16)
world_map
```

```python
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
```

```python
geolocator = Nominatim(user_agent="ny_explorer")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)
longitude=[]
latitude=[]

#address=['chennai','dallas','hsjhgdgh','London, UK']

for i in l1:
    location = geolocator.geocode(i)
    if location:
        latitude.append(location.latitude)
        longitude.append(location.longitude)
    else:
        latitude.append(np.NaN)
        longitude.append(np.NaN)
    
    
print(len(latitude)) 
```

```python
while np.NaN in latitude:
    latitude.remove(np.NaN)
while np.NaN in longitude:
    longitude.remove(np.NaN)
```

```python
len(latitude)
```

```python
len(longitude)
```

```python
incident_launchstations=folium.map.FeatureGroup()

for i , j in zip(latitude,longitude):
    incident_launchstations.add_child(
    folium.CircleMarker(
    [i,j],
    fill=True,
    fill_color='green',
    color='red',
    fill_opacity=0.6,
    radius=0.5
    )
    )
    
world_map.add_child(incident_launchstations)    
```
