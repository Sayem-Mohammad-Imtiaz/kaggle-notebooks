# Parte 1: Classificação Usando Palavras

```python
from sklearn.metrics import r2_score, classification_report, mean_squared_error, accuracy_score, confusion_matrix, roc_curve, roc_auc_score,auc
from sklearn.feature_extraction.text import HashingVectorizer, CountVectorizer
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as matplotlib


import pandas
import string
import numpy

%matplotlib inline
```

```python
# Andreis: Faça um teste com o Count Vectorizer

Teste = ["Esse review e bom",
          "Esse review e ruim",
          "Esta analise e fantástica",
          "Esta analise e ruim"]

Vectorizer = CountVectorizer()
Pos_palavras = Vectorizer.fit_transform(Teste)
print(Vectorizer.get_feature_names())
print(Pos_palavras.toarray())
```

```python
# Andreis: agora, mostra como funciona o transform

Nova_Review = Vectorizer.transform(["Esse review e novo"])
print(Vectorizer.get_feature_names())
print(Nova_Review.toarray())
```

```python
# Tarefa: Leia o CSV com o Pandas
# Dataset: Review = textos com uma review
#          Sentiment = sentimento (se é uma review positiva ou negativa)

Reviews = pandas.read_csv('../input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv')
print(Reviews.shape)
print(Reviews.head())
```

```python
# Tarefa: Divida o X e Y e use o train_test_split

x = Reviews['review'][0:5000].tolist()
y = Reviews['sentiment'][0:5000].tolist()
X_train, X_test, y_train, y_test = train_test_split(x, y)
```

```python
# Tarefa: Use o vetorizador CountVectorizer para fazer o fit_transform e transform
Vectorizer = CountVectorizer()
X_train_vectorized = Vectorizer.fit_transform(X_train) 
X_test_vectorized = Vectorizer.transform(X_test)
```

```python
# Tarefa: Use a regressão logística

model = LogisticRegression()
model.fit(X_train_vectorized,y_train)

result = model.predict(X_test_vectorized)
print(accuracy_score(result, y_test))
print(confusion_matrix(result, y_test))
```

```python
# Tarefa: Teste um exemplo negativo e um positivo

bad_review =  Vectorizer.transform(["This movie is bad"])
model.predict(bad_review)
```

```python
# Vamos descombrir por que deu o warning na célula passada?

from sklearn.model_selection import learning_curve
def plot_learning_curve(est, X_train, y_train) :
    train_sizes, train_scores, test_scores = learning_curve(estimator=est, X=X_train, y=y_train,cv=5,n_jobs=-1)
    train_mean = numpy.mean(train_scores, axis=1)
    train_std = numpy.std(train_scores, axis=1)
    test_mean = numpy.mean(test_scores, axis=1)
    test_std = numpy.std(test_scores, axis=1)
    matplotlib.figure(figsize=(8,10))
    matplotlib.plot(train_sizes, train_mean, color='blue', marker='o', markersize=5, label='training accuracy')
    matplotlib.fill_between(train_sizes, train_mean + train_std, train_mean - train_std, alpha=0.15, color='blue')
    matplotlib.plot(train_sizes, test_mean,color='green', linestyle='--',marker='s', markersize=5,label='validation accuracy')
    matplotlib.fill_between(train_sizes,test_mean + test_std,test_mean - test_std,alpha=0.15, color='green')
    matplotlib.grid(b='on')
    matplotlib.xlabel('Number of training samples')
    matplotlib.ylabel('Accuracy')
    matplotlib.legend(loc='lower right')
    matplotlib.show()

plot_learning_curve(model, X_train_vectorized, y_train)
```

# Parte 3: Regressão Linear

```python
# Função Reshape (relembrando da primeira aula prática)

def reshape(list1D):
     return numpy.array(list1D).reshape(-1,1)
```

```python
# Tarefa: Leia o dataset de salários com o Pandas

Salary = pandas.read_csv("../input/years-of-experience-and-salary-dataset/Salary_Data.csv")
Salary
```

```python
# Tarefa: divida x e y e faça um scatter com o matplotlib
x = Salary['YearsExperience']
y = Salary['Salary']
matplotlib.scatter(x,y)
```

```python
# Tarefa: Faça o train_test_split e use a regressão linear
x_train, x_test, y_train, y_test = train_test_split(x, y)

regression = LinearRegression()
regression.fit(reshape(x_train), y_train)
predict = regression.predict(reshape(x_test))
print(r2_score(predict,y_test))
```

```python
# Use o scatter plot para plotar os trains e tests com cores diferentes (use color='red')
matplotlib.scatter(x_train,y_train,color='blue')
matplotlib.scatter(x_test,y_test,color='red')
```

```python
# Tarefa: Pegue o coeficiente linear do modelo e onde ele intercepta o eixo y
coef = regression.coef_
b = regression.intercept_
print(coef, b)
```

```python
# Agora, usando numpy, construa uma função e plot junto do scatter plot (use o linspace para construir um espaço linear no x)
x = numpy.linspace(0,15,1000)
y = coef*x+b
matplotlib.scatter(x_train,y_train,color='blue')
matplotlib.scatter(x_test,y_test,color='red')
matplotlib.plot(x,y)
```
