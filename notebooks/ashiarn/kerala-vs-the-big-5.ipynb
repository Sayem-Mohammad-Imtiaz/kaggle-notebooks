Data set is obtained from kaggle, I can't vouch for its Accuracy. I appreciate the effort and time put in by the contributor



```python
df = pd.read_csv('/kaggle/input/covid19-corona-virus-india-dataset/complete.csv')
df =df[['Date','Name of State / UT','Cured/Discharged/Migrated','Death','Total Confirmed cases']]
df = df.rename(columns = {'Name of State / UT':"State",'Cured/Discharged/Migrated':"Cured",'Total Confirmed cases':"Total"})
df.head()
```

I had to make a new column storing the number of Active cases per date , so have to subtract the number of recovered and deceased entries.

```python
df['Active'] = df['Total']- (df['Cured']+df["Death"])
df.head()
```

```python
df_1 =df.pivot_table("Active",["State"],"Date")
df_1 = df_1.fillna(0)
df_1.head()
```

The pivot table has States in index, however to plot the data it is better to have states in columns 

```python

df_2 =df_1.iloc[:,30:] #selecting the dates from significant numbers starts reported
df_2 = df_2.transpose()
df_2.head()
```

```python
df_2.plot(kind="line",figsize=(20,10))
plt.legend(loc='upper left')
plt.show
```

The plot is cluttered so we are just looking at the top 5 states in india in terms of GDP and Kerala. (Kerala ranks 10th)

```python
df_3 = df_2[["Kerala","Maharashtra","Tamil Nadu","Gujarat","Karnataka","Uttar Pradesh"]]
df_3.head()
```

```python
df_3.plot(kind="line",figsize=(10,10))

plt.title("Kerala vs The Big 5")
plt.xlabel("Date")
plt.ylabel("Active cases")
```

It is evident that The curve of Kerala is **Flattening** compared to other states. Lets take a deeper Look

```python
df_3.iloc[20:32,:].plot(kind='line',figsize=(20,10))
```

By looking at the plot we see that at one point, Kerala had the **highest** number of cases in the country. The curve was going up and up. This was the same time when lock down started in the state one day before the country. So when did it change. For that lets look at recovery  curve of the same states


```python
df_4 =df.pivot_table("Cured",["State"],"Date")
df_4 = df_4.fillna(0)
df_4.head()
```

```python
df_5 = df_4.transpose()
df_5 = df_5[["Kerala","Maharashtra","Tamil Nadu","Gujarat","Karnataka","Uttar Pradesh"]].iloc[30:,:]
df_5.plot(kind='line',figsize=(20,10))
```

Even though it looks like Maharastra has the highest number of recovered patients. But its actually around  300 recovered out of 2500 where in kerala 200 recovered out of 380. **That is nearly 51%**. Clearly Kerala's investment in it's  health infrastructure  gave it the edge over other states. Ofcourse preventing the transmission with strict policing is also significant to control the spread.
