```python
#In this example I am using world happiness report to find out if freedom 
#to make life choices can predict GDP per capita of a country
# Outcome variable is "GDP PER CAPITA"
# Explanatory variable "Freedom to make life choices"

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import linear_model as lm


import os
print(os.listdir("../input/world-happiness-report-2021"))

d=pd.read_csv("../input/world-happiness-report-2021/world-happiness-report.csv")
d.head()
```

```python
d=d.rename(index=str,columns={"Log GDP per capita":"GDP"})
d = d[["GDP","Freedom to make life choices"]]
d.head()
```

```python
print(d.isnull().values.any())
print(d.isnull().sum())
```

```python
d = d.dropna()
print("Check for NaN/null values:\n",d.isnull().values.any())
print("Number of NaN/null values:\n",d.isnull().sum())
```

```python
# Simple Linear Regression 
inputDF = d[["Freedom to make life choices"]]
outcomeDF = d[["GDP"]]
model = lm.LinearRegression()
results = model.fit(inputDF,outcomeDF)

print(model.intercept_, model.coef_)
#print(model.summary())
```

```python
# it looks the predictor variable (Freedom to make life choices) has some influnce if not significat on the ouctome variable (GDP Per Capita)
yp2 = model.predict(inputDF)
print(yp2)

```

```python
plt.scatter(inputDF,outcomeDF)
plt.plot(inputDF,yp2, color="blue")
plt.xlabel("Freedom to make life choices")
plt.ylabel("GDP per capita")
plt.show()
```

```python
# Sum of square of residuals
print(np.sum(np.power(outcomeDF - yp2, 2)))
```
