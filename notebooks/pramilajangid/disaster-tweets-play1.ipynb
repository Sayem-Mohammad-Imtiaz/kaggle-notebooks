```python
def substitute0(i):
    i = re.sub("(?P<url>https?://[^\s]+)", " ", i)
    i = re.sub(r'(<.*?>+)', ' ', i)
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    i =  emoji_pattern.sub(r'', i)
    i = re.sub(r'[^A-Za-z0-9\.\,\'\"]', ' ', i)
    
    # i = re.sub(r'(-PRON-|PRON )', ' ', i)
    # i = re.sub(r'[^A-Za-z\s]', " ", i)
    
    i = re.sub(r'\s+', " ", i)
    return i
```

```python
with Pool() as p:
    positive_reviews_train = p.map(substitute0, positive_reviews_train)
    negative_reviews_train = p.map(substitute0, negative_reviews_train)
    
    
with Pool() as p:
    reviews_test = p.map(substitute0, reviews_test)
```

```python
# Let's remove stop words for now

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
sw = stopwords.words("english")
```

```python
sw
```

```python
sw.extend('a b c d e f g h i j k l m n o p q r s t u v w x y z etc'.split())
```

```python
sw
```

```python
def remove_stop_words(i):
    
    words = i.split()
    
    final_words = []
    
    for i in words:
        if i.lower() not in sw:

            final_words.append(i)
        
    return " ".join(final_words)
```

```python
with Pool() as p:
    positive_reviews_train = p.map(remove_stop_words, positive_reviews_train)
    
with Pool() as p:
    negative_reviews_train = p.map(remove_stop_words, negative_reviews_train)
```

```python
with Pool() as p:
    reviews_test = p.map(remove_stop_words, reviews_test)
```

```python
def substitute00(i):
    i = re.sub(r'[^A-Za-z\s\']', " ", i)
    i = re.sub(r'\s+', " ", i)
    doc = nlp(i)
    lemmatized_words = []
    for j in doc:
        lemmatized_words.append(j.lemma_)
    return (" ".join(lemmatized_words)).lower()
```

```python
with Pool() as p:
    positive_reviews_train = p.map(substitute00, positive_reviews_train)
    
    negative_reviews_train = p.map(substitute00, negative_reviews_train)
    
with Pool() as p:
    reviews_test = p.map(substitute00, reviews_test)
```

```python
from IPython.display import HTML
import pandas as pd
import numpy as np
import base64

# function that takes in a dataframe and creates a text link to  
# download it (will only work for files < 2MB or so)
def create_download_link(df, title = "Download CSV file", filename = "submission.csv"):  
    csv = df.to_csv(index=False)
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)



# create a link to download the dataframe
create_download_link(submission)
```
