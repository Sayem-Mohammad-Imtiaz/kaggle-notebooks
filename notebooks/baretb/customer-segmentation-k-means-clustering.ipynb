```python
#Gender Grouping
gender_group = df['Gender'].value_counts().to_frame()
gender_group = gender_group.reset_index()
gender_group = gender_group.rename(columns = {'index': 'gender', 'Gender': 'total'}, inplace = False)


#plot
fig = px.pie(gender_group, values='total', names='gender', title='Gender Ratio',hole=.3)
fig.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=False)
sns.set_style("white")

sns.regplot(ax=axes[0], x="Age", y="Annual Income (k$)", data=df)
sns.regplot(ax=axes[1], x="Age", y="Spending Score (1-100)", data=df)
sns.regplot(ax=axes[2], x="Spending Score (1-100)", y="Annual Income (k$)", data=df)
```

# K-means Clustering

### Selected Features: Age - Spending Score (1-100)

```python
from sklearn.cluster import KMeans

# Age - Spending Score (1-100)
X= df.iloc[:, [2,4]].values

wcss=[]
K = range(1,9)

for k in K: 
    kmeans = KMeans(n_clusters=k, init ='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
#Choosing the right number of clusters

plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")
ax = sns.lineplot(x=K,y=wcss)
ax.set(xlabel="cluster",ylabel="value", title="Elbow Graph")
```

```python
# Cluster number: 4

kmeans = KMeans(n_clusters=4, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_y = kmeans.fit_predict(X)

sns.set_style("white")
plt.figure(figsize=(12, 6))
sns.scatterplot(x=X[:,0], y=X[:,1], s=200, alpha=.7)
sns.scatterplot(x=kmeans.cluster_centers_[:, 0], y=kmeans.cluster_centers_[:, 1], s=600)
```
