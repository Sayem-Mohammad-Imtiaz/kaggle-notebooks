```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import keras
import sklearn

%matplotlib inline
pd.set_option('display.max_columns', None)
```

```python
data= pd.read_csv('../input/world-happiness/2019.csv')
df = pd.DataFrame(data)
```

```python
df['rank']=0
df['rank']= pd.cut(df['Overall rank'], bins=[0, 40, 78, 119, 157], labels = ['A', 'B', 'C', 'D'] )
```

```python
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)
```

```python
plt.figure(figsize=(11,7))
plt.plot('Overall rank',  # x
         'Generosity',  # y
         data=train_df, 
         linestyle='none', 
         marker='*', 
         markersize=20,
         color='green', 
         alpha=0.9)

plt.title('Overall rank of Generosity', fontsize=30)
plt.xlabel('Overall rank of Generosity', fontsize=25)
plt.ylabel('Generosity', fontsize=25)
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(11,7))
sns.boxplot(x="rank", 

            y="Perceptions of corruption", 

            data=train_df)

plt.show()
```

```python
x = ['Score', 'GDP per capita',
       'Social support', 'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']

x2 = [ 'GDP per capita',
       'Social support', 'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']

x_data = train_df.drop(['Score', 'Country or region', 'Overall rank', 'rank'], axis=1) 
x_data2 = train_df.drop(['Country or region', 'Overall rank', 'rank', 'Generosity', 'Score'], axis=1)
x_data3 = train_df.drop(['Country or region', 'Overall rank', 'rank', 'Generosity', 'Score', 'Perceptions of corruption'], axis=1)

y_data = train_df['rank']
```

```python
sns.pairplot(df, hue= "rank")
```

```python
x_test = test_df.drop(['Score', 'Country or region', 'Overall rank', 'rank'], axis=1) 
x_test2 = test_df.drop(['Score', 'Country or region', 'Overall rank', 'rank', 'Generosity'], axis=1)
x_test3 = test_df.drop(['Score', 'Country or region', 'Overall rank', 'rank', 'Generosity', 'Perceptions of corruption'], axis=1)

y_testdata = test_df['rank']
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier().fit(x_data,y_data)
print("훈련 세트 정확도:  {:.3f}.".format(dt.score(x_data, y_data)))
print("테스트 세트 정확도:  {:.3f}.".format(dt.score(x_test, y_testdata)))
print("특성 중요도:\n", dt.feature_importances_)
x_test
##depth 조절
dt = DecisionTreeClassifier(max_depth=3, random_state=0).fit(x_data,y_data)
print("훈련 세트 정확도:  {:.3f}.".format(dt.score(x_data, y_data)))
print("테스트 세트 정확도:  {:.3f}.".format(dt.score(x_test, y_testdata)))
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier().fit(x_data,y_data)

print("훈련 세트 정확도:  {:.3f}.".format(forest.score(x_data, y_data)))
print("테스트 세트 정확도:  {:.3f}.".format(forest.score(x_test, y_testdata)))
print("특성 중요도:\n", forest.feature_importances_)

forest2 = RandomForestClassifier().fit(x_data3,y_data)

print("훈련 세트 정확도:  {:.3f}.".format(forest2.score(x_data3, y_data)))
print("테스트 세트 정확도:  {:.3f}.".format(forest2.score(x_test3, y_testdata)))
print("특성 중요도:\n", forest2.feature_importances_)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3).fit(x_data,y_data)

print("훈련 세트 정확도:  {:.3f}.".format(knn.score(x_data, y_data)))
print("테스트 세트 정확도:  {:.3f}.".format(knn.score(x_test, y_testdata)))

knn2 = KNeighborsClassifier(n_neighbors=3).fit(x_data2,y_data)

print("훈련 세트 정확도:  {:.3f}.".format(knn2.score(x_data2, y_data)))
print("테스트 세트 정확도:  {:.3f}.".format(knn2.score(x_test2, y_testdata)))
```
