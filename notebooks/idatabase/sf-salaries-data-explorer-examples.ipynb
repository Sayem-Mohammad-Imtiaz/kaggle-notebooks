```python
import pandas as pd
```

Read Salaries.csv as a dataframe

```python
sal = pd.read_csv('../input/Salaries.csv',dtype={"Overtimdtype": object})
```

Preview data

```python
sal.head()
```

Find out how many entries there are

```python
sal.info()
```

What is the average BasePay ?

```python
pd.to_numeric(sal['BasePay'], errors='coerce').dropna().mean()
```

What is the highest amount of Overtime Pay ?

```python
pd.to_numeric(sal['OvertimePay'], errors='coerce').dropna().max()
```

What is the name of highest paid person ?

```python
sal.loc[sal['TotalPayBenefits'].idxmax()]['EmployeeName']
```

What is the name of lowest paid person ?

```python
sal.loc[sal['TotalPayBenefits'].idxmin()]['EmployeeName']
```

What was the average paid(TotalPayBenefits) of all employees per year ?

```python
sal.groupby('Year').mean()['TotalPayBenefits']
```

How many job titles are there?

```python
sal['JobTitle'].nunique()
```

What are the most common jobs?

```python
sal['JobTitle'].value_counts().head(1)
```

How many people have the word Chief in their job title?

```python
sum(sal['JobTitle'].apply(lambda x: 'chief' in x.lower()))
```
