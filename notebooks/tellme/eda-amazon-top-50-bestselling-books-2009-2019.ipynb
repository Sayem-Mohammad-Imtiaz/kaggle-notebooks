# Number of Reviews

### Duplicate Review count numbers!
From the Review data below, we can see that there are duplicate review numbers.  

```python
total_reviews_by_author = df.groupby(by=['Author','Genre','Name','Year'])['Reviews'].sum().reset_index().sort_values(by=['Reviews'],ascending=False).reset_index(drop=True)
total_reviews_by_author.head(10)
```

Upon further investigation(visually inspecting) we can determine that whenever a book is on the list more than once, the User Rating and Price can change, but the Review number does not

```python
pd.set_option('display.max_rows', None)
duplicate = df[df.duplicated(['Genre','Name', 'Author'],keep=False)]
duplicate.sort_values(by=['Genre','Name', 'Author'],ascending=False)
```

Non Fiction books seem to have a longer shelf life than Fiction; particularly self-help books  

```python
df_check_title = df['Name'].value_counts().reset_index()
df_check_title.columns = ['Name', 'Yearly Top 50 count']
df_check_title = df_check_title.merge(df[['Name','Genre']].drop_duplicates(), on='Name', how='left')

g = sns.catplot(y='Name',x='Yearly Top 50 count', hue='Genre',
                data=df_check_title.head(25),
                kind='bar', height=6, aspect=2)
plt.title('Occurances in top 50 from 2009-2019') 
plt.xlabel('Count')
plt.ylabel('Title')
plt.show()
```

# User Ratings

```python
g = sns.catplot(x='User Rating', hue='Genre', 
                data=df,
                kind='count', height=6, aspect=2)
plt.title('Count of Genre User Ratings') 
plt.show()
```

```python
plt.figure(figsize=(14,6))
plt.title("Average User Ratings by Year") 
g = sns.pointplot(x="Year", y="User Rating",color=pal[5],
                   data=df, dodge=True)
plt.figure(figsize=(14,6))
plt.title("Average Genre User Ratings by Year") 
g = sns.pointplot(x="Year", y="User Rating", hue="Genre",
                   data=df, dodge=True)
```

# Prices

```python
plt.figure(figsize=(14,6))
plt.title("Average Price by Year") 
g = sns.pointplot(x="Year", y="Price", color=pal[5],
                   data=df, height=10,aspect=2.5, dodge=True)

plt.figure(figsize=(14,6))
plt.title("Average Genre Price by Year") 
g = sns.pointplot(x="Year", y="Price", hue="Genre",
                   data=df, height=10,aspect=2.5, dodge=True)
```
