# **BUILDING DATA FRAMES FROM SCRATCH**

For building a new data frame:
*     I create four different lists(1st one is for the name of books, 2nd one is for the author's names, 3rd one is for the number of pages and 4th one is for the label of columns).
*     Using first three lists i create a new list which includes this lists values. This list would be data frame's columns.
*     With **zip()** i zipped the label list and the column list. And assigned the result to a dictionary.
*     Then, i turned this dictionary to a data frame with Pandas's **DataFrame()** method.

```python
name = ["Gorulmeyenler", "Eastern Turkey"]
author = ["Roy Jacobsen", "Sevan Nisanyan"]
pages = [179,195]
ListLabel = ["name","author","pages"]
ListColumn = [name,author,pages]
zipp = dict(zip(ListLabel,ListColumn))
df = pd.DataFrame(zipp)
df
```

You can add a new column which includes different or same values to data frame. 
If you apply this with different values, you should write all rows one by one. 
If you give same value to entire column, create a column and just type the value. It is called as broadcasting.

```python
df["type"] = ["Novel","Travel"] #add new column
df["year"] = 0 #broadcasting
df
```

# **INDEXING PANDAS TIME SERIES**


If you want to add a date column to your data set, you may create a date's list. Of course, these dates type will be object(string). So, you must convert their types to  "datetime" with Pandas's **to_datetime()** method. 

```python
time_list = ["1994-09-07","1986-01-06"]
print(time_list)
print(type(time_list[1]))
time_list = pd.to_datetime(time_list)
print(time_list)
print(type(time_list))
```

Here, your data set turned into a time series.

```python
import warnings
warnings.filterwarnings("ignore")

data1 = data.head()
date_list = ["2000-01-10","2000-01-02","2000-03-11","2000-04-05","2001-07-08"]
date_list = pd.to_datetime(date_list)
data1["date"] = date_list
data1 = data1.set_index("date")
data1
```

Now, you can filter your data set according to date value.

```python
print(data1.loc["2000-01-10"])
```

```python
print(data1.loc["2000-01-01" : "2000-03-03"])
```

# **RESAMPLING PANDAS TIME SERIES**

With **resample()** you can create a new sample from your data set. 

```python
data1.resample("A").mean()
```

In this example we resampled our data set by year. And then, calculated the average of each year.

You can use "M" parameter for resample by month:

```python
data1.resample("M").mean()
```

With interpolation, you can fill the missing values. But of course, just numeric values.

```python
data1.resample("M").first().interpolate("linear")
```

```python
data1.resample("M").mean().interpolate("linear")
```
