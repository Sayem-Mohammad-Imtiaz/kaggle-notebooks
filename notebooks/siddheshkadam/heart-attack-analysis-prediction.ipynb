# Pie chart
- Age :Age of the patient 
- cp :Chest Pain type chest pain type
    - Value 1: typical angina
    - Value 2: atypical angina
    - Value 3: non-anginal pain
    - Value 4: asymptomatic
- fbs : (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
- restecg : resting electrocardiographic results
    - Value 0: normal
    - Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
    - Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
- exng :exercise induced angina (1 = yes; 0 = no)
- caa : number of major vessels (0-3)

```python
fig,ax=plt.subplots(2,3)
plt.subplots_adjust(left=0,right=2,bottom=2,top=4,wspace=0.4,hspace=0.4)

ax[0,0].set_title("Age of the patient")
ax[0,0].pie(age_count,labels=label_age,autopct='%1.2f%%')

ax[0,1].set_title("Chest Pain type chest pain type")
ax[0,1].pie(cp_list,labels=label_cp,autopct='%1.2f%%')

ax[1,0].set_title("fasting blood sugar > 120 mg/dl (1 = true; 0 = false)")
ax[1,0].pie(fbs_list,labels=label_fbs,autopct='%1.2f%%')

ax[1,1].set_title("resting electrocardiographic results")
ax[1,1].pie(restecg_list,labels=label_restecg,autopct='%1.2f%%')

ax[0,2].set_title(" exercise induced angina (1 = yes; 0 = no)")
ax[0,2].pie(exng_list,labels=label_exng,autopct='%1.2f%%')

ax[1,2].set_title("number of major vessels (0-3)")
ax[1,2].pie(caa_list,labels=label_caa,autopct='%1.2f%%')
```

# Splitting data

```python
train=heart.iloc[:,:-1]
print(train.shape)
```

```python
test=heart.iloc[:,-1:]
print(test.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(train,test,test_size=0.3,random_state=100)
```

```python
print("The size of x_train is :",x_train.shape)
print("The size of x_test is :",x_test.shape)
print("The size of y_train is : ",y_train.shape)
print("The size of y_test is : ",y_test.shape)
```

# Performing Standard Scaling 

```python
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

# Performing Logistic Regression and prediction

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
y_pred
```

# Accuracy check

```python
accuracy_score(y_pred,y_test)
```
