# Time based Analysis of Drake songs
Based on the above DataFrame i have selected the top 3 songs,
that are **Hotline Bling**, **God's Plan** and **Controlla**

 I will be plotting each songs with respect to the ranks they had with the passing of the dates. 

```python
hotline_bling= df[(df['artist']=='Drake') & (df['song']=='Hotline Bling')][['date','rank']]
godsplan= df[(df['artist']=='Drake') & (df['song']=="God's Plan")][['date','rank']]
controlla= df[(df['artist']=='Drake') & (df['song']=="Controlla")][['date','rank']]
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 8))

hotline_bling.plot(ax=axes[0],x='date',y='rank',label='hotline bling')
plt.xlabel('Date')
plt.ylabel('rank')
#plt.title('hotline bling')



godsplan.plot(ax=axes[1],x='date',y='rank',label='gods plan', color='r')
plt.xlabel('Date')
plt.ylabel('rank')
#plt.title('gods plan')



controlla.plot(ax=axes[2],x='date',y='rank',label='controlla', color='g')
plt.xlabel('Date')
plt.ylabel('rank')
#plt.title('controlla')
```
