## COVID Vaccination Programme Analysis

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import cufflinks as cf
cf.go_offline()
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
import plotly.express as px
```

```python
# Importing Data
df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv')
df.head()
```

```python
# Grouping by Country and Filling null values to 0 and Sorting the dataset based on number of people_vaccinated
dfc = df.groupby(['country']).max().sort_values('people_vaccinated',ascending=False).fillna(value=0)
dfc.head()
```

#### Vaccine Usage By Countries

```python
dfc['people_vaccinated'].iplot(kind='bar',bins=30,colors='blue',xTitle='iso_code',yTitle='Total no. of people vaccinated')
#dfw.iplot(kind='bar',x='iso_code',y='people_vaccinated',colors='blue',xTitle='Countries_codes',yTitle='People Vaccinated')
```

#### Top 5 Countries with Fastest rate of Daily vaccinations in the world (ppm)
This is interesting because it is w.r.t the population of countries and can be a little misleading

```python
dfc.nlargest(5,'daily_vaccinations_per_million')['daily_vaccinations_per_million']
```

```python
dfch = df[['country','iso_code','people_vaccinated','daily_vaccinations','daily_vaccinations_per_million','vaccines']]
dfch.fillna(0).head()
```

```python
dfw = dfch.groupby('country',as_index=False).max().sort_values('people_vaccinated', ascending=False)
dfw.fillna(0).head()
```

```python
fig = px.choropleth(locations = dfw['iso_code'],color=dfw['people_vaccinated'],
                    title=' World COVID vaccination - Number of People Vaccinated',
                    color_continuous_scale='plasma_r',
                   projection = 'equirectangular',
                   scope = 'world')
fig.show()
```
