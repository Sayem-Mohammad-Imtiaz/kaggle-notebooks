```python
data = pd.read_csv('../input/Dataset_spine.csv')
print (data)
```

```python
data=data.rename(columns={'Col1':'pelvic incidence', 'Col2':'pelvic tilt', 'Col3': 'lumbar lordosis angle', 'Col4':'sacral slope', 'Col5': 'pelvic radius', 'Col6':'degree spondylolisthesis', 'Col7': 'pelvic slope',  'Col8': 'Direct tilt' , 'Col9':'thoracic slope' , 'Col10': 'cervical tilt' , 'Col11':'sacrum angle' , 'Col12':'scoliosis slope'  })
print (data)
```

```python
data = data.as_matrix()
```

```python
data=data[:,:-1]
print (data)
```

```python
X=data[:,:-1]
print (X)
```

```python
Y=data[:,-1]
print (Y)
```

```python
sel=np.random.permutation(Y.shape[0])
print (sel)
```

```python
n70=int(Y.shape[0]*0.7)
Ytrain=Y[sel[:n70]]
print (Ytrain)
```

```python
Xtrain=X[sel[:n70],:]
print (Xtrain)
```

```python
Ytest=Y[sel[n70:]]
print (Ytest)
```

```python
Xtest=X[sel[n70:],:]
print (Xtest)
```

```python
from sklearn.svm import SVC 
```

```python
clf=SVC(C=1)
```

```python
clf.fit(Xtrain,Ytrain)
```

```python
clf.score(Xtest,Ytest)
```
