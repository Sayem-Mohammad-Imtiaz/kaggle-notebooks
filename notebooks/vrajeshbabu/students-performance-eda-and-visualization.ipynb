# Gender

```python
gender=df.groupby(["gender"])["math score","reading score","writing score"].mean()
#gender=gender.to_frame()
gender=gender.reset_index()
gender
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
gender.plot(x="gender",y=["math score","reading score","writing score"],kind="bar",color=['purple','orange','brown'],figsize=(8,4),width=0.2)
plt.xlabel("Gender",size=15)
plt.ylabel("Marks",size=15)
plt.title("Marks in each category",size=20)
plt.show()
```

```python
#Exporting it to a csv file
gender.to_csv("Gender.csv",index=False)
```

# Gender with race/ethnicity

```python
genderwithrace=df.groupby(["race/ethnicity"])["math score","reading score","writing score"].mean()
genderwithrace=genderwithrace.reset_index()
```

```python
genderwithrace
```

```python
genderwithrace.plot(x="race/ethnicity",y=["math score","reading score","writing score"],kind="bar",color=['purple','orange','brown'],figsize=(10,8))
plt.xlabel("Gender",size=15)
plt.ylabel("Marks",size=15)
plt.title("Marks in each category",size=20)
plt.show()
```

```python
race=df.groupby(["gender","race/ethnicity"])["math score","reading score","writing score"].mean()
race
```

```python
race = race.reset_index()
race
```

```python
male = race[race['gender']=='male']
male
```

```python
male.plot(x="race/ethnicity",y=["math score","reading score","writing score"],kind="bar",color=['purple','orange','brown'],figsize=(10,5))
plt.xlabel("Groups",size=15)
plt.ylabel("Marks",size=15)
plt.title("Marks of Male candidates in each rage",size=20)
plt.legend()
plt.show()
```

```python
female = race[race['gender']=='female']
female
```

```python
female.plot(x="race/ethnicity",y=["math score","reading score","writing score"],kind="bar",figsize=(10,5))
plt.xlabel("Groups",size=15)
plt.ylabel("Marks",size=15)
plt.title("Marks of Female candidates in each race",size=20)
plt.legend()
plt.show()
```

```python
#Exporting all the files to csv
race.to_csv("raceethnicity.csv",index=False)
male.to_csv("male.csv",index=False)
female.to_csv("female.csv",index=False)
```

## parental level of education

```python
ple=df.groupby(["gender","parental level of education"]).count()
ple = ple.reset_index()
ple
```

```python
ple=ple[['gender','parental level of education','race/ethnicity']]
ple.rename(columns={"race/ethnicity":"count"},inplace=True)
```

```python
maleple=ple[ple['gender']=='male']
femaleple=ple[ple['gender']=='female']
```

```python
figure = plt.figure(figsize=(10,8))
plt.pie(maleple["count"],labels=maleple["parental level of education"])
plt.title("Parental level for Male students",size=20)
plt.show()
```

```python
figure = plt.figure(figsize=(10,8))
plt.pie(femaleple["count"],labels=femaleple["parental level of education"])
plt.title("Parental level for Female students",size=20)
plt.show()
```

```python
#Exporting to csv

maleple.to_csv("male parental.csv",index=False)
femaleple.to_csv("femal parental.csv",index=False)
```

## Lunch

```python
lunch=df.groupby(['lunch']).count()
lunch.reset_index(inplace=True)
lunch
```

```python
lunch = lunch[['lunch','gender']]
lunch.rename(columns={"gender":"count"},inplace=True)
lunch
```

```python
figure = plt.figure(figsize=(8,8))
plt.pie(lunch["count"],labels=lunch["lunch"],colors=["yellow","orange"])
plt.title("Lunch status of children",size=20)
plt.show()
```

```python
#Exporting it to csv
lunch.to_csv("Lunch.csv",index=False)
```
