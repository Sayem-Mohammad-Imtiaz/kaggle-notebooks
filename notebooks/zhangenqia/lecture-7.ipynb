# create a global map for covid vaccinations

```python

import numpy as np 
import pandas as pd 
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df.index = pd.to_datetime(df['date'])
df=df.sort_index()
print(df.columns)
fig=px.choropleth(df,locations='country',locationmode='country names',color='daily_vaccinations',hover_name='country',animation_frame='date',color_continuous_scale='rainbow',range_color=(0.0,100.0))
fig.update_layout(title_text="daily_vaccinations",title_x=0.35)
fig.show()

plotly.offline.plot(fig,filename='vaccinations.html')
```

as mentioned in the info of the dataset,Daily vaccinations means for a certain data entry, the number of vaccination for that date/country,I choose it to show daily trend.With range for 0~100,we can see more details about it(unfortunately most of area is totally red)

# **create a global map of earthquake occurrences worldwide**

```python
df = pd.read_csv('../input/202105-recent-earthquake/query.csv',header=0)
df.index = pd.to_datetime(df['time'])
df['time']=df.index.strftime('%Y-%m-%d %H:00:00')
df = df.sort_index()
fig=px.scatter_geo(df,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='rainbow',range_color=(1.0,3.0))
fig.show()
```

I noticed that alaska suffered from earthquake frequently,so I made a research about it.For the source news, the earthquake just happened at 4.8.It also talked about the causes of the earthquake which is because of the subducting Pacific Plate,it is a smaller sibling of the Anchorage Earthquake ,although even deeper than the Anchorage Earthquake.
