**Preprocessing**

---



---



```python
df = pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
```

```python
feat = df.columns
col = (df[feat]==0).sum()
print(col)
```

```python
df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
```

```python
df.isnull().sum()
```

**Data Vizualization**

---



---



```python
msno.matrix(df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']], figsize=(12,8))
plt.grid()
```

```python
df['Glucose'].fillna(df['Glucose'].median(), inplace=True)
df['BloodPressure'].fillna(df['BloodPressure'].median(), inplace=True)
df['BMI'].fillna(df['BMI'].median(), inplace=True)
```

```python
Glucose_Age_Insulin = df.groupby(['Glucose'])
def fillna_insulin(series):
    return series.fillna(series.median())
df['Insulin'] = Glucose_Age_Insulin['Insulin'].transform(fillna_insulin)
df['Insulin'] = df['Insulin'].fillna(df['Insulin'].mean())
```

```python
df['SkinThickness'].fillna(df['SkinThickness'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.isnull().values.any()
```

**Standard Scaling**

---



---



```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)
```

```python
from sklearn.metrics import confusion_matrix,auc,accuracy_score,classification_report,roc_curve
```

**SVM Model**

---



---



```python
from sklearn.svm import SVC
svmm = SVC(kernel='rbf')
svmm.fit(xtrain,ytrain)
```

```python
ypred = svmm.predict(xtest)
accuracy_score(ytest,ypred)
```

```python
print(classification_report(ytest,ypred))
```

```python
fpr, tpr,_=roc_curve(ytest,ypred)
roc_auc = auc(fpr,tpr)
plt.style.use("dark_background")
plt.figure()
plt.plot(fpr,tpr,label='ROC Curve(area=%2.f)'%roc_auc)
plt.plot([0,1],[0,1],'k--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.title("ROC Curve")
plt.legend(loc='lower right')
```

**Random Forest Model**

---



---



```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(xtrain,ytrain)
```

```python
Ypred = classifier.predict(xtest)
accuracy_score(ytest,Ypred)
```

```python
print(classification_report(ytest,Ypred))
```

```python
fpr, tpr,_=roc_curve(ytest,Ypred)
roc_auc = auc(fpr,tpr)
plt.style.use("dark_background")
plt.figure()
plt.plot(fpr,tpr,label='ROC Curve(area=%2.f)'%roc_auc)
plt.plot([0,1],[0,1],'k--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.title("ROC Curve")
plt.legend(loc='lower right')
```

**KNN Model**

---



---



```python
from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors=3)
clf.fit(xtrain,ytrain)
print(clf.score(xtest,ytest))
```

```python
Yypred = clf.predict(xtest)
print(classification_report(ytest,Yypred))
```

```python
fpr, tpr,_=roc_curve(ytest,Yypred)
roc_auc = auc(fpr,tpr)
plt.style.use("dark_background")
plt.figure()
plt.plot(fpr,tpr,label='ROC Curve(area=%2.f)'%roc_auc)
plt.plot([0,1],[0,1],'k--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel("False positive rate")
plt.ylabel("True positive rate")
plt.title("ROC Curve")
plt.legend(loc='lower right')
```
