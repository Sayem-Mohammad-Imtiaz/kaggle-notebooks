**Introduction to Python**

Matplotlib: - This is a python library allow us to plot data. Different type of plots are as below:

    1. Line plot: - It is used if X axis is time.
    2. Scatter plot: - Shows the relationship between two variables.
    3. Histogram: - It is used to see the distribution of the data.

```python
import matplotlib.pyplot as plt

#line plot
data.Speed.plot(kind='line',color = 'g',label = 'Speed',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
data.Defense.plot(kind='line',color = 'g',label = 'Defense',linewidth=1,alpha = 0.5,grid = True,linestyle = '-')
plt.legend(loc='upper right')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('line plot')
plt.show()
```

```python
#scatter plot
data.plot(kind='scatter',x='Attack',y='Defense',alpha=0.5,color='red')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Scatter plot')
plt.show()
```

```python
#histogram
data.hist(column='Speed',bins=50,figsize=(12,12))
```

**While and for loops**

```python
i=0
while(i !=5):
    print('i is:',i)
    i+=1
print(i,'is equal to 5')
```

```python
#global and local scope
x=2
def f():
    x=3
    return x
print(x)
print(f())
```

```python
#nested function
def square():
    def add():
        x=2
        y=4
        z=x+y
        return z
    return add()**2
print(square())
```

```python
#default and flexible arguments
def f(a,b=2,c=3):     #default
    y=a+b+c
    return y
print(f(1))
print(f(1,4,3))
```

```python
#flexible arguments
def f(*args):
    for i in args:
        print(i)
f(1)
print('---------')
f(1,2,3)
```

```python
def f(**kwargs):
    for key,value in kwargs.items():
        print(key,'', value)
f(country='India',capital='New Delhi')
```

```python
#lambda function
square= lambda x: x**2
square(5)
```
