# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tpot import TPOTClassifier
```

# Load dataset

```python
df = pd.read_csv("../input/heart-attack-analysis-prediction-dataset/heart.csv")
df.sample(10)
```

```python
df.info()
```

# Quick analysis

```python
from pandas_profiling import ProfileReport
profile = ProfileReport(df, title='Heart Attack Data Analysis Report')
profile.to_widgets()
```

# Split the data set

```python
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=44)
```

```python
print(X_train.shape)
print(X_test.shape)
```

# Define and perform the search for best model

```python
tpot_model = TPOTClassifier(generations=5, population_size=20, cv=10, random_state=44, verbosity=2)
tpot_model.fit(X_train, y_train)
```

# evaluate the best model


```python
print(tpot_model.score(X_test, y_test))
```

# export the corresponding Python code for the optimized pipeline to a text file

```python
tpot_model.export('tpot_exported_pipeline.py')
```
