```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import resample
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
import sklearn.metrics as metrs
import matplotlib.pyplot as plt

data=pd.read_csv('//kaggle//input//credit-card-customers//BankChurners.csv')

data.head(10)



```

```python

#-----------------Data QC and Visulization
print(data.dtypes)



#----------------------check how many unknow in the data
print(data['Education_Level'].value_counts(),data['Marital_Status'].value_counts())

#-----------------
lst=(data.select_dtypes('object').columns)
fig,axes=plt.subplots(2,3,figsize=(8,6),gridspec_kw=dict(hspace=1,wspace=1))
for i,ax in enumerate(axes.flat):
    ax.bar((data[lst[i]].value_counts()).index,data[lst[i]].value_counts())
    ax.set_title(lst[i])
    ax.set_xticklabels((data[lst[i]].value_counts()).index,rotation=45, ha='right')
```

```python
#-------------------------Data Selection and prepration
data['y']=(data['Attrition_Flag']).replace(['Existing Customer','Attrited Customer'],[0,1])
y=data['y']
#------------------------------------select x data
X=data.iloc[:,2:-3]
X.columns
#-----------------------change the features of the categories to numerical(encoding)
Xencoding=pd.get_dummies(X,columns=['Gender', 'Dependent_count', 'Education_Level','Marital_Status', 'Income_Category', 'Card_Category'])
Xencoding.columns


#--------------------scale the data
Xscaling=StandardScaler().fit_transform(Xencoding)
```

```python

#-----------------split the data for train and test
Xtrain,Xtest,ytrain,ytest=train_test_split(Xscaling,y,random_state=42)
print(ytrain.shape)
print(Xtrain.shape)
print(ytest.shape)
print(Xtest.shape)


```

```python
#---------------------------Model an QC
gb=GradientBoostingClassifier().fit(Xtrain,ytrain)
print('Test score-----------',gb.score(Xtest,ytest))
print('Train score-----------',gb.score(Xtrain,ytrain))

#-----------------------Model QC
metrs.plot_confusion_matrix(gb,Xtest,ytest,display_labels=['neg','pos'])
metrs.plot_roc_curve(gb,Xtest,ytest)

ypred=gb.predict(Xtest)
print('Accuracy-----',metrs.accuracy_score(ytest,ypred))
```
