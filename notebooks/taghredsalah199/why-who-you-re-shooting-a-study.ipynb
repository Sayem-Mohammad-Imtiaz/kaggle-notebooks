## 1- Importing Libraries & Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df_police= pd.read_csv('../input/data-police-shootings/fatal-police-shootings-data.csv')
df_police
```

```python
df_police.info()
```

## 2- Discribe the Race with manner of death and the age

```python
plt.figure(figsize=(20,10))
sns.boxplot(x="race", y="age",hue="manner_of_death", data=df_police, palette="coolwarm")
```

### The W race are the most shotted, O rase are the most shot and tasered!

## 3- Describe the sign of mental illness with gender and age:

```python
plt.figure(figsize=(15,10))
sns.violinplot(x="gender", y="age",hue="signs_of_mental_illness", data=df_police,palette='rainbow')
```

## 4- Describe threat level with age and race :

```python
df_police['race']=df_police['race'].astype('string')
```

```python
plt.figure(figsize=(15,10))
sns.boxenplot(x="race", y="age",hue="threat_level", data=df_police)
```

## 5- Distribution of age :

```python
plt.figure(figsize=(15,10))
sns.distplot(df_police['age'])
```

## The most repeated ages are from 30 to 40!

## 6- Describe flee with the age

```python
plt.figure(figsize=(15,10))
sns.boxenplot(x="flee", y="age", data=df_police,palette='Set1')
```

## We can say the most of them not fleeing!

## 7- Describe if camera catch body with gender and age

```python
plt.figure(figsize=(15,10))

sns.stripplot(x="body_camera", y="age", data=df_police,hue='gender',jitter=True,palette='Set1')
```

## 8- Describe threat level with gender and age

```python
plt.figure(figsize=(15,10))

sns.stripplot(x="threat_level", y="age", data=df_police,hue='gender',jitter=True)
```

## Threat level is attack mostly!
