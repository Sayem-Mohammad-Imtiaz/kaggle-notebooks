> ## import data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from sklearn.model_selection import train_test_split

from sklearn import metrics

from sklearn.linear_model import LinearRegression

%matplotlib inline
```

## import data

```python
df=pd.read_csv('../input/real-estate-price-prediction/Real estate.csv')
```

## X and Y is primary and final features

```python
x=df.drop(['Y house price of unit area','No'], axis=1)

y=df['Y house price of unit area']
```

## Split data to train and test data

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

## define model and train model 

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
pd.DataFrame(list(model.coef_),index=x.columns)
```

## Predict model with test data

```python
y_pred = model.predict(X_test)
```

## Calculate the residuals

```python
mae=metrics.mean_absolute_error(y_test,y_pred)
mse=metrics.mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
ind=['mae','mse','rmse']
pd.DataFrame([mae,mse,rmse],index=ind)
```
