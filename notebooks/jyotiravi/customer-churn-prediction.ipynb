```python
#Import all necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
# Read cvs file
data = pd.read_csv('../input/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv')
```

```python
data.describe
```

```python
#Read first 5 line of the csv file
data.head()
```

```python
#Build a linear model where dependant variable is 'monthly_charges' and independant variable is 'tenure'
#data = pd.read_csv('../input/telco-customer-churn')
x= data.loc[:,['tenure']].values
y= data.loc[:,['MonthlyCharges']].values
```

```python
# Devide the dataset into train and Test datasets in 70 & 30 ratio
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
#Build the model on train set and pridect the values on test set
from sklearn.linear_model import LinearRegression
simpleLinearRegression = LinearRegression()
simpleLinearRegression.fit(x_train,y_train)
```

```python
#After predicting the values, find the Root Mean Square error
y_pred=simpleLinearRegression.predict(x_test)
y_pred
```

```python
#Compare with above predicted values with the actual values
y_test
```

```python
#Root mean square
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_pred, y_test)
rmse=np.sqrt(mse)
rmse
```

```python
#rmse value is high so this Linear regression model is not good. if rmse value is less than 10 we can consider this model
```

```python
#1. Build the simple logistic regression model where dependant or target variable is "Churn" & 
# independant variable is 'Monthly charges'
y= data.loc[:,['Churn']].values
x= data.loc[:,['MonthlyCharges']].values
```

```python
y_test
```

```python
#Multiple Logistic regression
#1. build the multiple logistic regression model where dependant variable is 'Churn' and
#  independant variable are 'tenure' & 'MonthlyCharges'
# i. Devide dataset in 80:20 ratio
#ii. Build the model on train set and predict the values on test set
#iii. Build the confusion matrix and get the accuracy score
```

```python
y= data.loc[:,['Churn']].values
x= data.loc[:,['MonthlyCharges','tenure']].values
```

```python
# i. Devide dataset in 80:20 ratio
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.80, random_state=0)
```

```python
#ii. Build the model on train set and predict the values on test set
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
y_pred = logmodel.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
```

```python
#Decision tree
#Build the decision tree model where dependent variable is 'Churn' and Independant variable is 'Tenure'
#i. Devide dataset in 80:20 ratio
#ii. Build the midel on train set and predict the values on test set
#iii. Build the confusion matrix and calculate the accuracy score
```

```python
#Build the decision tree model where dependent variable is 'Churn' and Independant variable is 'Tenure'
x=data.loc[:,['tenure']].values
y=data.loc[:,['Churn']].values
```

```python
y_test
```
