```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

A look at the data

```python
df = pd.read_csv('/kaggle/input/all-trumps-twitter-insults-20152021/trump_insult_tweets_2014_to_2021.csv')
```

```python
df.head()
```

```python
df.isna().sum()
```

Clean up, fill the null target columns with `nothing`

```python
df.target = df.target.fillna("nothing")
```

Create a word cloud to quick view the major phrases/words

```python
stopwords = set(STOPWORDS)
text = df.tweet.to_csv() # the actual text containing the tweets
wc  = WordCloud(
    background_color='white',
    stopwords=stopwords
)
wc.generate(text)
```

Using matplotlib to show the wordcloud as image

```python
plt.figure(figsize=(14,14))
plt.imshow(wc)
plt.show()
```
