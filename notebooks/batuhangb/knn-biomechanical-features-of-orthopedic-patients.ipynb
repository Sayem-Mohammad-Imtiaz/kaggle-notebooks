```python
data = pd.read_csv("../input/biomechanical-features-of-orthopedic-patients/column_3C_weka.csv")
data.info()
```

```python
data.head()
```

```python
plt.scatter(data["pelvic_incidence"],data["pelvic_radius"],color = "green",alpha=0.3)
plt.scatter(data["pelvic_tilt"],data["pelvic_radius"],color = "red",alpha=0.3)
plt.show()
```

```python
data["class"] = [0 if each == "Normal" else 1 if each == "Spondylolisthesis" else 2 for each in data['class']]
```

```python
x_data = data.drop(["class"],axis = 1)
y = data["class"].values
```

```python
# Normalization
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data))
x.head()
```

```python
# train test split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.3,random_state= 1)
```

```python
# find k 
score_list= []

for each in range(1,50):
    knn = KNeighborsClassifier(n_neighbors = each)
    knn.fit(x_train,y_train)
    score_list.append(knn.score(x_test,y_test))

    
```

```python
plt.plot(range(1,50),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```

```python
# k = 43

knn = KNeighborsClassifier(n_neighbors = 43)
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)

print("Score:{}".format(knn.score(x_test,y_test)))
```
