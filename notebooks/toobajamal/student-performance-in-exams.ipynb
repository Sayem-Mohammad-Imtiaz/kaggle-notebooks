```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "../input/students-performance-in-exams/StudentsPerformance.csv"
student_df=pd.read_csv(url)
student_df.head()
```

```python
student_df.tail()
```

```python
student_df.isna().sum()
```

```python
student_df.describe(include= "all")
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (15, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
math_perf = student_df[['gender', 'math score']]
plt.figure(figsize=(8,8))
plt.title('Gender wise exam performance in math')
sns.barplot(x=math_perf['gender'], y=math_perf['math score']);
```

```python
reading_perf = student_df[['gender', 'reading score']]
plt.figure(figsize=(8,8))
plt.title('Gender wise exam performance in reading')
sns.barplot(x=reading_perf['gender'], y=reading_perf['reading score']);
```

```python
writing_perf = student_df[['gender', 'writing score']]
plt.figure(figsize=(8,8))
plt.title('Gender wise exam performance in writing')
sns.barplot(x=writing_perf['gender'], y=writing_perf['writing score']);
```

```python
ethnicity = student_df[['race/ethnicity', 'math score', 'reading score', 'writing score']]
ethnicity_df = ethnicity.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean()
ethnicity_df
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1, 3, 1)
plt.xticks(rotation=75)
plt.title('Maths')
sns.barplot(x=ethnicity_df.index, y=ethnicity_df['math score']);

plt.subplot(1, 3, 2)
plt.xticks(rotation=75)
plt.title('Reading')
plt.tight_layout(pad = 2)
sns.barplot(x=ethnicity_df.index, y=ethnicity_df['reading score']);

plt.subplot(1, 3, 3)
plt.xticks(rotation=75)
plt.title('Writing')
plt.tight_layout(pad = 2)
sns.barplot(x=ethnicity_df.index, y=ethnicity_df['writing score']);
plt.suptitle('Ethnicity wise exam performance')
```

```python

parent_edu_df = student_df[['parental level of education', 'math score', 'reading score', 'writing score']].groupby('parental level of education')[['math score', 'reading score', 'writing score']].mean()
parent_edu_df
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1, 3, 1)
plt.xticks(rotation=75)
plt.title('Maths')
sns.barplot(x=parent_edu_df.index, y=parent_edu_df['math score']);

plt.subplot(1, 3, 2)
plt.xticks(rotation=75)
plt.title('Reading')
plt.tight_layout(pad = 2)
sns.barplot(x=parent_edu_df.index, y=parent_edu_df['reading score']);

plt.subplot(1, 3, 3)
plt.xticks(rotation=75)
plt.title('Writing')
plt.tight_layout(pad = 2)
sns.barplot(x=parent_edu_df.index, y=parent_edu_df['writing score']);
plt.suptitle("Does parents' education have effects on student exam performance?")
```

```python
lunch_df = student_df[['lunch', 'math score', 'reading score', 'writing score']].groupby('lunch')[['math score', 'reading score', 'writing score']].mean()
lunch_df
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1, 3, 1)
plt.xticks(rotation=75)
plt.title('Maths')
sns.barplot(x=lunch_df.index, y=lunch_df['math score']);

plt.subplot(1, 3, 2)
plt.xticks(rotation=75)
plt.title('Reading')
plt.tight_layout(pad = 2)
sns.barplot(x=lunch_df.index, y=lunch_df['reading score']);

plt.subplot(1, 3, 3)
plt.xticks(rotation=75)
plt.title('Writing')
plt.tight_layout(pad = 2)
sns.barplot(x=lunch_df.index, y=lunch_df['writing score']);
plt.suptitle("Does lunch type effects student performance?")
```

## Conclusion
 - Since all the marks are out of 100, the mean values are the percentage values as well
 - Male students perform better in maths exam while female students are good at reading and writing
 - E ethnic group performs best in exams and A performs the worst of all
 - Students who have educated parents perform well. Higher the educated parents, higher the students perform
 - Students who have standard lunch perform well as compared to those who get free or reduced lunch
 - Students who prepare for the exams perform better
