```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import plotly.express as px
%matplotlib inline
```

```python
df = pd.read_csv("../input/pima-indians-diabetes-database/diabetes.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
df[['Glucose', 'BMI', 'BloodPressure', "Insulin", "SkinThickness"]] = df[['Glucose', 'BMI', 'BloodPressure', "Insulin", "SkinThickness"]].replace(0, np.nan)
df.isnull().sum()
```

```python
for col in ['Glucose','BMI', 'BloodPressure', 'Insulin', 'SkinThickness']:
    df[col] = df[col].fillna(df[col].mean())
```

```python
df.describe()
```

```python
is_diabetic = {0: 'No', 1: 'Yes'}
df['is_diabetic'] = df.Outcome.map(is_diabetic)
```

## Pregnancies

```python
fig = px.histogram(df, 
                   x='Pregnancies', 
                   marginal='box', 
                   color='is_diabetic',
                   nbins=18, 
                   title='Distribution of Pregnancies')
fig.update_layout(bargap=0.1)
fig.show()
```

The distribution is declining as the number of pregnancies increases. Highest number of pregnancy is 1 with 135 women and least is 16 with zero women. Less number of women are diabetic in each pregnancy range but all of them are diabetic in pregnancies greater than 13.

## Glucose

```python
df.Glucose.describe()
```

Minimum glucose level is 44 and maximum is 199

```python
fig = px.histogram(df, 
                   x='Glucose', 
                   marginal='box', 
                   color='is_diabetic',
                   nbins=100, 
                   title='Distribution of Glucose')
fig.update_layout(bargap=0.1)
fig.show()
```

Highest glucose level is 100-101.9 and least is 44 with almost normal distribution and it also shows that diabetic people have more glucose

```python
fig = px.histogram(df, 
                   x='BloodPressure', 
                   marginal='box', 
                   color='is_diabetic',
                   title='Distribution of BloodPressure')
fig.update_layout(bargap=0.1)
fig.show()
```

Distribution of Blood Pressure is normal as well with highest of 70-79

```python
fig = px.histogram(df, 
                   x='SkinThickness', 
                   marginal='box', 
                   color='is_diabetic',
                   nbins=50, 
                   title='Distribution of Skin Thickness')
fig.update_layout(bargap=0.1)
fig.show()
```

```python
corr = df.corr()
sns.heatmap(corr, annot=True);
```

Glucose has highest correlation with outcome meaning people with higher glucose levels likely to have diabetes

## Conclusion
 - Women with pregnancies higher than 13 are diabetic
 - Women with higher BMI are more likely to be diabetic
 - Women with higher age are more diabetic
 - Most common insulin level among all individuals is 150-199 with 440 women out of which 39.09% are diabetic
 - Glucose has highest impact on whether a person is diabetic or not
