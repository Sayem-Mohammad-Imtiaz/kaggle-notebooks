## Cleaning the Data

```python
df.columns = df.columns.str.upper() #Changing all column names to uppercase
df.head(5)
```

```python
df.set_index('NAME',inplace = True) #Set the column 'NAME' as index
df.head(5)
```

```python
df.index = df.index.str.replace('.*(?=Mega)','') #Eliminating the data prefix
df.head(5)
```

## Exploratory Analysis

### **Some questions...**

**Which Pokemon has the biggest attack?**

```python
df[df['ATTACK']== df['ATTACK'].max()]
```

**Which Pokemon has the lowest defense?**

```python
df[df['DEFENSE']== df['DEFENSE'].min()]
```

**How many legendary and non_legendary Pokemon are there?**

```python
df['LEGENDARY'].value_counts()
```

Different type of Pokemon in TYPE1

```python
print('These are the types of Pokemon in TYPE 1:', df['TYPE 1'].unique())
print('\n')
print('This the total type of Pokemon:', df['TYPE 1'].nunique())
print('\n')
print('The count for each type is: \n', df['TYPE 1'].value_counts())
```

Countplot of TYPE 1

```python
plt.figure(figsize = (14,8))
sns.countplot(x = df['TYPE 1'], data = df) #Water is the most common type in TYPE 1
```

Different type of Pokemon in TYPE 2:

```python
print('These are the types of Pokemon in TYPE 2:', df['TYPE 2'].unique())
print('\n')
print('This the total type of Pokemon:', df['TYPE 2'].nunique())
print('\n')
print('The count for each type is:\n', df['TYPE 2'].value_counts())
```

Countplot of TYPE 2

```python
plt.figure(figsize = (14,8))
sns.countplot(x = df['TYPE 2'], data = df) #Flying is the most common type in TYPE 2
```

### Type 1 and 2 combined

```python
df['TYPE 1-2'] = df['TYPE 1']+ '-' + df['TYPE 2']
df.head()
```

### What are the biggest combinations between the types of Pokemon?

```python
print('The biggest combinations between the types of Pokemon are:\n', df['TYPE 1-2'].value_counts().head(10))
print('\n')
print('The number of combinations in Pokemons are: ',df['TYPE 1-2'].nunique() )


plt.figure(figsize = (12,8))
sns.countplot(x = 'TYPE 1-2',data = df, order = df['TYPE 1-2'].value_counts().iloc[:10].index) #Counplot of the biggest combinations
```

Normal-Flying is biggest combination

### Strongest legendaries Pokemon

```python
legendary = df[df['LEGENDARY']==True].reset_index()
strong_legendary = legendary.sort_values(by='TOTAL',ascending = False).head(5)

plt.figure(figsize=(12,8))
sns.barplot(x = 'NAME', y ='TOTAL', data = strong_legendary, palette = 'viridis')
plt.title('Strongest Legendary Pokemons')
```

### Strongest non-legendaries Pokemon

```python
non_legendary = df[df['LEGENDARY']==False].reset_index()
strong_non_legendary = non_legendary.sort_values(by='TOTAL',ascending = False).head(5)

plt.figure(figsize=(12,8))
sns.barplot(x = 'NAME', y ='TOTAL', data = strong_non_legendary, palette = 'viridis')
plt.title('Strongest Non-Legendary Pokemons')
```

### Strongest and weakest types

```python
plt.figure(figsize = (12,8))
sns.boxplot(x = 'TYPE 1', y = 'TOTAL', data = df)
```

We can see that the strongest type of pokemon is Dragon and the type do avoid is Bug

### Pokemon type that has more legendary and the type that doesn't have legendary

```python
plt.figure(figsize = (12,8))
sns.countplot(x = 'TYPE 1',data = legendary, order = df['TYPE 1'].value_counts().iloc[:18].index)
```

As we can see the type that has most legendary is Pyschic and the types that dont'have legendary are Bug, Poison and Fighting in TYPE 1

```python
plt.figure(figsize = (12,8))
sns.countplot(x = 'TYPE 2',data = legendary, order = df['TYPE 2'].value_counts().iloc[:18].index)
```

As we can see, the type in TYPE 2 that has most legendary is Flying and the types that don't have legendaries are Poison, Grass, Rock and Bug.

So the Types that don't have legendaries are Poison and Bug.
