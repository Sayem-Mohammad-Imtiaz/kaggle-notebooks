### These are the statistic Results

```python
print(df.count())
df['title'].value_counts().count()
```

* The total number of entries of books are 11,123.
* But due to the repeated entries of some books the total number of books is not correct. Therefore, the unique count is given as  10,348.

The ratings' count gives us an idea about how popular is that particular book. The top five books reviewed by people are
1. Twilight (Twilight #1)	
2. The Hobbit or There and Back Again
3. The Catcher in the Rye	
4. Angels & Demons (Robert Langdon #1)	
5. Harry Potter and the Prisoner of Azkaban 


```python
df.sort_values('ratings_count', ascending= False, inplace=True, na_position='first')
df.head()
```

```python
text_reviews = df['text_reviews_count']
label = df['title']
index = np.arange(len(text_reviews))
```

```python
# plotting purpose
plt.barh(label, index)
plt.ylabel('Books')
plt.xlabel("Ratings' count")
plt.title('Text review count of the books')
plt.show()
```
