```python
colors = ('red', 'orange', 'lightcyan', 'yellow', 'teal', 
          'palegreen', 'mistyrose', 'olive', 'lavender', 'dodgerblue', 'royalblue', 'purple', 'pink', 'cyan')
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot() 

ex = np.zeros(len(colors))
ex[-4:]=(1.,0.7,1.,0.5)

# <Stay>s 
ax.pie(stay,
       autopct='%1.1f%%',
       textprops={'size': 12},
       colors = colors, pctdistance=0.8, radius=1.3,
       startangle=90, counterclock=False, labeldistance=1.2,explode=ex)
ax.set_title('<Stay>s', fontsize=30, loc='left')
plt.subplots_adjust(left=0.0, bottom=0.0, right=0.85)

# use legend
plt.legend(stay.index.unique(), loc="center right", fontsize=12, 
           bbox_to_anchor=(1.35, 0.5), bbox_transform=plt.gcf().transFigure)
plt.show()
```
