```python
#necessary as the later use of io seems to overwrites the default path
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename=='charts.csv':
            path = os.path.join(dirname, filename)
        else:
            pass
```

```python
import numpy as np
import pandas as pd

from PIL import Image

from wordcloud import WordCloud,STOPWORDS
from matplotlib import pyplot as plt

import requests
import io
```

```python
#prepair song words count
ds = pd.read_csv(path)
songs = pd.Series(ds.song.str.split(" "))

sw = [w.capitalize() for w in list(STOPWORDS)]
slist_ = pd.Series([st for row in songs for st in row]).value_counts()

slist = slist_[~slist_.index.isin(sw)].to_dict()
```

```python
response = requests.get("https://raw.githubusercontent.com/ambader/ambader/main/MISC/tune_t.png")
tune_mask = np.array(Image.open(io.BytesIO(response.content)))

plt.figure(figsize= (15,32))
wordcloud = WordCloud(colormap="Pastel1",max_words=2000,width = 200, height = 500, background_color ="lavender",min_font_size = 1, mask=tune_mask, prefer_horizontal=0.5).generate_from_frequencies(slist)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
