```python
train_x = train.drop(['POSTED_BY', 'BHK_OR_RK',
      'READY_TO_MOVE', 'RESALE', 'ADDRESS', 'LONGITUDE',
       'LATITUDE', 'TARGET(PRICE_IN_LACS)'], axis = 'columns')
```

```python
test_x = test.drop(['POSTED_BY', 'BHK_OR_RK',
      'READY_TO_MOVE', 'RESALE', 'ADDRESS', 'LONGITUDE',
       'LATITUDE'], axis = 'columns')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(train_x, target)
```

```python
test_x['TARGET(PRICE_IN_LACS)'] = predict
```
