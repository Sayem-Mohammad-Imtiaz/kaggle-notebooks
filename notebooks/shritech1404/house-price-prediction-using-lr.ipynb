```python
X = house.drop(["id", "date", "floors", "waterfront", "long", "lat", "price"], axis="columns")
```

```python
X
```

```python
y = house.price
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
model.predict([[5, 6.00, 1190, 5250, 0, 3, 6, 2180, 0, 1955, 0, 98178, 1340, 5650]])
```
