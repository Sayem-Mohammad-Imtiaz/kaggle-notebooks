**Average BasePay (Remove the values 'Not provided' from BasePay, typecast the column to float and find mean**

**Top 5 most common jobs**

```python
sal['JobTitle'].value_counts().head(5)
```

**How many have the word chief in their job title**

```python
def find_chief(job_title):
    count=0
    for title in job_title:
        if('CHIEF' in title.upper().split()):
            count+=1
    return count

find_chief(sal['JobTitle'])
```

**Top 5 average paid jobs**

```python
sal.groupby('JobTitle').mean()['TotalPayBenefits'].sort_values(ascending=False).head(5)
```
