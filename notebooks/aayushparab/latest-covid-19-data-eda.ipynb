```python

s=df[['State/UTs','Total Cases','Active','Discharged','Deaths']].sort_values(by='Total Cases',ascending=False)
s.set_index('State/UTs',inplace=True)
s.head(7)
```

```python
#Total Cases
fig = px.choropleth(
    df,
    geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
    featureidkey='properties.ST_NM',
    locations='State/UTs',
    color='Total Cases',
    color_continuous_scale='Reds',
    title='Total Cases In Different States-'
)


fig.update_geos(fitbounds="locations", visible=False)

fig.show()
```

The above graph clearly shows about the extent to which Maharashtra is affected as compared to other states in the country

```python
#The discharge rate is the most important among any other categories since it 
#determines the state's ability to cure them as quickly as possible.

fig = px.choropleth(
    df,
    geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
    featureidkey='properties.ST_NM',
    locations='State/UTs',
    color='Discharge Ratio (%)',
    color_continuous_scale='Reds',
    title='Discharge Ratio In Different States-'
)


fig.update_geos(fitbounds="locations", visible=False)

fig.show()
```

The above map shows the discharge rates in different states and we can infer that the the states in the eastern part of India have relatively lesser discharge rate which implies that the medical facilities there need to be improved.
