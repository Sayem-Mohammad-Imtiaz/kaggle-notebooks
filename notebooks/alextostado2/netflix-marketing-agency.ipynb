# Realizando las primeros queries o estructuras

```python
# Revisar la sinopsis de una pelicula
synopsis = netflix['description']
synopsis[1956]
```

```python
# Analizar las peliculas segun su año de publicación
netflix.release_year.describe()
```

```python
# Dataframe de 'TV Show'
tv_shows = netflix[netflix.type == 'TV Show']
tv_shows.sample(5)
```

```python
# Ajustar variables para procesamiento
tv_shows_split = tv_shows['duration'].str.split(' ', expand = True)
tv_shows_split.columns = ['Seasons', 'Null']
tv_shows_split
type(tv_shows_split['Seasons'])
tv_shows_split.describe()
tv_shows_split.sample(10)
```

```python
# Busqueda de Cobra Kai en dataframe de Netflix
cobra_kai = netflix[netflix.title == 'Cobra Kai']
cobra_kai
```

```python
# Segmentar las peliculas de año de 2017 o posteriores, percentil 50-100%
titles_2017 = netflix.release_year >= 2017
titles_2017 = netflix[titles_2017]
```

```python
# Segmentar el percentil 0-50%
titles_down2017 = netflix[netflix.release_year < 2017]
titles_down2017.head(3)
```

```python
# Revisión del codigo anterior
titles_2017.describe()
```

```python
#Revisión del codigo anterior
titles_2017.tail(3)
```

```python
# Ordenar los valores por el año de lanzamiento
netflix.sort_values('release_year', ascending=False).head(5)
```

```python
# Agrupando el numero de peliculas y directores por año

netflix_anual = netflix.groupby('release_year')[['title']].count()
netflix_anual
```

```python
# Agrupando el numero de peliculas por pais

netflix_country = netflix.groupby('country')[['title']].count()
netflix_country
```

```python
# Ordenando los paises con mayor numero de peliculas publicadas

netflix_country.sort_values('title', ascending = False).head(10)
```

```python
# Buscando a mi director favorito 'Martin Scorsese'

titles_scorsese = netflix[netflix.director == 'Martin Scorsese']
titles_scorsese.head(10).sort_values('duration')
```

```python
# Buscando a mi director favorito 'Martin Scorsese'

titles_drama = netflix[netflix.release_year == 2002]
titles_drama.head(2)
```

# Visualizando los primeros datos

```python
# Importar librerias de visualización de datos

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
# Numero de peliculas y directores por año
sns.set_style("darkgrid")

plt.figure(figsize = (16, 8))

plt.plot(netflix_anual, marker = 'x')

plt.xlabel('Year')
plt.ylabel('Number of movies')

plt.title('Peliculas lanzadas por año')
plt.legend('Movies')
```

```python
#Histograma de peliculas

plt.figure(figsize = (16, 8))
plt.hist(netflix.release_year, bins = 30);
```
