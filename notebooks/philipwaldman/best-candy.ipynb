# Import Libraries

```python
import pandas as pd
from sklearn import linear_model
```

# Load the Data

```python
df = pd.read_csv('../input/the-ultimate-halloween-candy-power-ranking/candy-data.csv')
df.head()
```

# Fit the Linear Model

```python
features = ['chocolate', 'fruity', 'caramel', 'peanutyalmondy', 'nougat', 'crispedricewafer', 'hard', 'bar', 'pluribus']
X = df[features]
y = df['winpercent']
reg = linear_model.LinearRegression()
reg.fit(X, y)
inter = reg.intercept_
coefs = reg.coef_
score = reg.score(X, y)
score
```

# Candy Score Equation

```python
equation = 'winpercent ='
for feature, coef in zip(features, coefs):
    if coef >= 0:
        equation += ' + '
    else:
        equation += ' - '
        coef *= -1
    equation += f'{round(coef, 3)} * {feature}'
if inter >= 0:
    equation += f' + {round(inter, 3)}'
else:
    inter *= -1
    equation += f' - {round(inter, 3)}'
equation
```

# Table of Weights

```python
df_weights = pd.DataFrame({'feature': features, 'weight': coefs})
df_weights.sort_values(by=['weight'], ascending=False, inplace=True)
df_weights
```
