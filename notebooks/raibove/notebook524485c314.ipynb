```python
#check which coulmn contain na values
nan_values = data.isna()
nan_values.any()
```

```python
# If suicide column has na values replace with zero
data['suicides_no'] = data['suicides_no'].fillna(0)

nan_values = data.isna()
nan_values.any()
```

```python
data.dropna(inplace=True)
data.isna().any()
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
sns.set_style("whitegrid")

country_df = data.groupby('year')[['suicides_no','sex']].sum()
plt.plot('suicides_no','o--r',data=country_df)
plt.ylabel('Suicide rate')
plt.xlabel('Year')
plt.title("Total Suicides each year")
```

```python
sns.stripplot(x='sex', y='suicides_no', data=data, jitter=True, edgecolor='none', alpha=.40)
sns.despine()
```

```python
plt.figure(figsize=(15,5))
sns.stripplot(x='age',y='suicides_no',data=data,hue='sex',dodge=True,jitter=True,edgecolor='none', size=8.5,alpha=0.25)
```

```python
plt.figure(figsize=(25,8))
sns.stripplot(x='year',y='suicides_no',data=data,hue='sex',dodge=True,jitter=True,edgecolor='none',alpha=0.25)
```

```python
plt.figure(figsize=(10,8))
res = data.loc[data['country'] == "Albania"]
sns.scatterplot('year','suicides_no',hue="age",data=res)
```

```python
sns.set_style("whitegrid")

country_df = data.groupby('country')[['suicides_no']].sum()
plt.xticks(rotation=90)
plt.tick_params(axis='x', which='major', labelsize=3)
plt.plot('suicides_no','o--r',data=country_df)
plt.ylabel('Suicide rate')
plt.xlabel('Country')
plt.title("Total Suicides for every country")
```
