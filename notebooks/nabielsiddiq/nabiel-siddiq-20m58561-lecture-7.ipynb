# Global map indicating vaccination progress
The code below is intended to show the progress of vaccinations of Covid-19 among countries worldwide.
The output of the code is global time-animation map.

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

#import data from kaggle dataset
data1=pd.read_csv('../input/d/gpreda/covid-world-vaccination-progress/country_vaccinations.csv',header=0)

#re-indexing dataframe using "date" column
data1.index=pd.to_datetime(data1['date'])
data1=data1.sort_index()

#visualization (total population vaccinated per hundred)
fig_cv = px.choropleth(data1,locations='country',locationmode='country names',color='people_fully_vaccinated_per_hundred',hover_name='country',animation_frame='date',color_continuous_scale='Turbo',range_color=(0.0,50.0))
fig_cv.update_layout(title_text="Total Number of Population Have Received 2 Doses of Vaccines (in %)",title_x=0.45)
fig_cv.show()
#exporting maps into html (total population vaccinated per hundred)
plotly.offline.plot(fig_cv,filename='Fully Vaccinated Map.html')

#visualization (daily vaccinations per million)
fig_dv = px.choropleth(data1,locations='country',locationmode='country names',color='daily_vaccinations_per_million',hover_name='country',animation_frame='date',color_continuous_scale='Turbo',range_color=(0.0,10000.0))
fig_dv.update_layout(title_text="Daily Vaccinations per Million",title_x=0.45)
fig_dv.show()
#exporting maps into html
plotly.offline.plot(fig_dv,filename='Daily Vaccinated per Million Map.html')
```

The first map shows the total number of populations that have received 2 doses of vaccines (deemed as "fully vaccinated"). The second map provides a more complete data in terms of countries as compared to the first one (there are data of some african countries for this variable), thus giving more complete view of how countries are progressing in terms of vaccinations in daily terms. Some countries (such as United States, Morocco, United Kingdom, etc.) are seen to be improving the daily vaccinations as the time progress.

# Global map indicating earthquake events

A global map showing earthquake events with magnitude of > 6 Richter Scale in the Pacific Rim since the beginning of 2021 up to recent date.

```python
import pandas as pd
import numpy as np
import plotly.express as px

data1=pd.read_csv('../input/pacirif-rim-eq/Pacific rim - 2021.csv')
data1.index=pd.to_datetime(data1['time'])
data1['time']=data1.index.strftime('%Y-%m-%d %H:00:00')
data1=data1.sort_index()

fig_eq=px.scatter_geo(data1,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='rainbow',range_color=(1.0,8.0))
fig_eq.update_geos(
    resolution=50,
    showcoastlines=True, coastlinecolor="RebeccaPurple",
    showland=True, landcolor="White",
    showocean=True, oceancolor="LightBlue",
)
fig_eq.update_layout(title_text="Earthquake Event (Mag > 6 Richter Scale) in the Pacific Rim - 2021",title_x=0.5)
fig_eq.show()
```

According to USGS Earthquake Catalog, there are 57 earthquake events with magnitude more than 6 Richter Scale happened in the Pacific Rim region since the year 2021 began. The region is selected because it is considered as one of the most active tectonic region which generates a lot of earthquakes.

Throughout 2021, there has been no major earthquake disaster happened in the region. However, based on the catalog, there is a huge earthquake, with magnitude of about 8.3 richter scale, occurred in the Kermadec Island, northern part of New Zealand. Fortunately, there is no inhabitant on the island, thus no casualties resulted from the event. Several damages did happen in the island since the eartquake caused around 300 landslides, which in overall reform parts of the natural landscapes there.

The eartquake actually generated a small tsunami that affect surrounding islands, even up to Australia (64 cm), Fiji (13 cm), and Tonga (6 cm). However, it was so small that it did not affect any damage to the affected areas.
