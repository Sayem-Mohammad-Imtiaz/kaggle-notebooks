```python
X=data.drop(columns=['DEATH_EVENT'])
y=data['DEATH_EVENT']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=101)
```

**Using XGBoost classifier**

```python
import xgboost as xgb
```

```python
xgb=xgb.XGBClassifier()
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators=[int(x) for x in np.linspace(start=100,stop=1500,num=10)]
max_features=['auto', 'sqrt', 'log2']
max_depth=[int(x) for x in np.linspace(10,100,10)]
min_samples_split=[2,4,6,10,14]
min_samples_leaf=[3,5,7,13,19]

```

```python
random_grid={'n_estimators':n_estimators,
             'max_features':max_features,
             'max_depth':max_depth,
             'min_samples_split' :min_samples_split,
             'min_samples_leaf':min_samples_leaf,
             'criterion':['entropy','gini']}
print(random_grid)
```

```python
xgb_rv=RandomizedSearchCV(estimator=xgb,param_distributions=random_grid,n_iter=100,n_jobs=-1,random_state=101)
```

```python
model=xgb_rv.fit(X_train,y_train)
```

```python
xgb_rv.best_params_
```

```python
best_random_grid=xgb_rv.best_estimator_
```

```python
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
y_pred=best_random_grid.predict(X_test)
print(confusion_matrix(y_test,y_pred))
print("Accuracy Score {}".format(accuracy_score(y_test,y_pred)))
print("Classification report: {}".format(classification_report(y_test,y_pred)))
```

**ANN**

```python
#load the dependencies
import tensorflow
from tensorflow import keras
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout,BatchNormalization
```

```python
model=Sequential()
#first dense layer
model.add(Dense(32,activation='relu',input_dim=12))
model.add(BatchNormalization())
# second layer
model.add(Dense(32,activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
#o/p layer
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='nadam',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=32,batch_size=8,validation_data=(X_test,y_test))
```

```python
model.evaluate(X_test,y_test)
```
