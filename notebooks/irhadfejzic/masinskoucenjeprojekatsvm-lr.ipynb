## Metoda potpornih vektora (SVM) i logistička regresija

<div style="text-align: justify; margin-bottom: 10px;">
Metoda potpornih vektora, kao i logistička regresija, je klasifikacioni model koja koristi pristup nadglednog obučavanja. Oba algoritma su zasnovana na treniranju i učenju kroz date podatke kako bi bili adekvatni za testiranje i predikciju/klasifikaciju. Nadgledno obučavanje je metoda koja doprinosi treniranju mašina/metoda za klasifikaciju objekata, problemi i situacija zavisno od relacionih ulaznih podataka. Metode kao SVM i logistička regresija primaju karakteristike, dimenzije, boje, uzorke, visine, itd. dok mašine nisu u stanju da prezicno vrše klasifikaciju. U ovom projektu, u većini slučajeva, treniranje će zauzimati 80% podataka iz dataset-a, dok ostatak će se vršiti na testiranje i provjeravanje preciznosti.
</div>
<div style="text-align: justify; margin-bottom: 10px;">
Logistička regresija je algoritam koji se koristi u rješavanju problema klasifikacije. Koristi prediktivnu analizu da opiše podatke i objasni vezu između varijabla. Logistička regresija se vrši na ulaznu varijablu A gdje je izlazna varijabla B diskretne vrijednosti od 0 do 1 (false ili true). 
Metoda potpornih vektora je model koji se koristi za klasifikaciju kao i rješavanje regresionih problema (izlazna varijabla je realna ili neprekidna vrijednost kao "težina ili plata"). Algoritam izvršava postavljanje hiperravni ili prave (decision boundary) koja odvaja podatke u klase. 
</div>
<div style="text-align: justify; margin-bottom: 10px;">   
SVM pokušava da nađe najbolju marginu (udaljenost između prave i potpornih vektora) za razdvajanje podataka u klase i samim tim umanjivanjem greške pogrešne klasifikacije. U ovoj metodi biramo potporne vektore, kroz dataset za treniranje, koji imaju najveću snagu kod definiranja hiperravni. Logistička regresija pronalazi pravu, za odvajanje klasa, minimizirajući neku klasifikacijsku grešku (u algoritmu to će biti funkcija troška). Koristi sigmoidnu funkciju koja raspoređuje podatke u 2 klase zavisno da li je vrijednost tog podatka >= 50% ili ne. U ovom projektu vršit ćemo analiziranje obje metode kroz razne vrste podataka, a zatim i korištenjem već implementiranih mašina kroz biblioteku sklearn, vidjet ćemo kako funkcionišu i malo kompleksnije vrste algoritama.
</div>

## Opis metode potpornih vektora

<div style="text-align: justify">
U mašinskom učenju, metoda potpornih vektora spada u metode nadgledanog obučavanja gdje algoritam uči iz datih primjera/podataka o kojim se već zna ishod. Kod metoda potpornih vektora (SVM) glavni zadatak algoritma je da pronađe korektnu pravu (ili hiperravan) u N-dimenzionalnom prostoru koja dijeli date podatke po osobinama. Ovakav algoritam je veoma koristan kod klasifikacije raznih podataka poput životinjskih čelija, ukoliko tražimo kancerogene čelije na velikim podacima, klasificiranja studenata po njihovim ocjenama, klasifikacija emailova, detekcije lica, itd. Nalaženje takve hiperravni vrši se posmatranjem maksimalne margine, tj. maksimalne udaljenosti između podataka raličitih klasa. Ovo omogućuje da budući podaci pripradaju korektnoj klasi iako postoji malo odstupanje od ostalih podataka te klase. Nalaženje takve hiperravni izvršava se pomoću određivanja potpornih vektora (support vectors), tački koje su bliže hiperravni i imaju uticaj na poziciju i orijentaciju hiperravni. Preko potpornih vektora maksimiziramo marginu koja razdvaja 2 klase.

Sama dimenzija hiperravni ovisi o broju karakteristika datih podataka. Ukoliko se podaci sastoje samo od dvije ulazne karakteristike tada je hiperravan data u obliku prave. Za veću dimenziju hiperravan je u obliku dvodimenizionalne ravni. Nalaženje otpimalne prave radimo preko jednačine prave date sa y=ax+b. Tako da za date vektore W (težinski vektor) i x (set tačaka) iste dimenzije, zapis hiperravni možemo prikazati u sljedećoj formi:
hw(x) = w <sup>T</sup>x+b = 0, gdje je b bias (slobodan član).
SVM algoritam nastoji maksimizirati marginu između skupa podataka i hiperravni. Funkcija troška pomaže kod maksimiziranja takve margine koju nazivamo "hinge loss" funkcija predstavljena kao: 
    <br/><br/>
    <font size="5px">L(w)= <span>&Sigma;</span>max(0,1-y<sup>i</sup>[w <sup>T</sup>x+b]) + λ||w||<sub>2</sub><sup>2</sup></font>
    <br/><br/>
Prvi dio funkcije, koji se naziva hinge loss, kažnjava greške loše klasifikacije dok drugi dio je regularizacija, tehnika izbjegavanja preprilagodavanja tako što kažnjava velike koeficijente u vektoru rješenja. Cilj je minimizirati prethodnu funkciju za nalaženje makimalne margine. Ukoliko se dati podatak nalazi tačno na margini rezultat funkcije će biti 0.<br/><br/>

```python
#Biblioteke koje se koriste
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches 
import random
import math

#Klasa dijeli dataset na 80% za treniranje i ostatak za testiranje
class MetodaPotpornihVektora(object):
    """
    Inicijalizacija, postavljanje boja i grafa
    """
    def __init__(self, dataA, dataB, crtanje=True,  prikaziSve=True):
        self.dataA = dataA
        self.dataB = dataB
        self.crtanje = crtanje
        self.prikaziSve = prikaziSve
        self.boje = {1:'r',-1:'b'}
        #Postavljanje grafa za 1 red, 1 kolonu (Ukoliko postoji više podgrafa)
        if self.crtanje:
            self.fig = plt.figure(figsize=(15,10))
            self.ax = self.fig.add_subplot(221)
            
    def postavi_podatke(self, dataA, dataB):
        """
        Postavljanje podataka u posebni niz koji drži tip i vrijednosti
        """
        for i in range(len(dataA)):
            dataA[i].insert(0, 1)
        for i in range(len(dataB)):
            dataB[i].insert(0, 0)

        data = dataA + dataB

        for i in range(len(data)):
            data[i].insert(0, 1)
            data[i] += [data[i].pop(1)]
            if data[i][3] == 1:
                data[i].append(-1)
            else:
                data[i].append(1)
            data[i].pop(3)
            data[i] = np.array(data[i])
            
        random.shuffle(data)
        return np.array(data)

    def odvoji_podatke(self, data_final):
        """
        Odvajanje Podataka
        """
        #Koristi se 80% za treniranje algoritma, dok je ostatak korišten za testiranje i gledanje preciznosti
        postotak80 = int((80/100 *len(data_final)))

        skup_testiranja = data_final[postotak80:]
        skup_treniranja = data_final[:postotak80]
        return skup_treniranja, skup_testiranja

    def train(self, Epoha = 100000):
        """
        Funkcija koja trenira algoritam kroz epohe. Generalno je postavljeno na 100000 epoha, dok zavisno kolika je veličina algoritama može se i smanjiti
        """
        #Funkcija preuzima broj epoha i niz podataka koje će koristiti za treniranje
        self.dataset = self.postavi_podatke(self.dataA, self.dataB)
        self.skup_treniranja, self.skup_testiranja = self.odvoji_podatke(self.dataset)
        
        skup_treniranja = self.skup_treniranja
        skup_testiranja = self.skup_testiranja
        
        w = np.array([0,0,0])
        stopa = 1
        konvergira = False
        k = 0
        min_distance_negative = 0 #nalaženje najmanje udaljenosti za marginu
        min_distance_positive = 0
        while(konvergira == False) and k < Epoha: #algoritam radi dok ne završi sa epohama ili dok ne nađe optimalnu hiperravan

            k +=1
            nasumican_vektor = int(random.uniform(0,len(skup_treniranja)-1))
            x_i = skup_treniranja[nasumican_vektor][:-1]
            y_i = skup_treniranja[nasumican_vektor][-1]

            n_t = 1/(k*stopa)

            if (y_i*np.dot(x_i,w)) < 1: # Ukoliko je y*(<w,x>) <1 moramo ažurirati vektor težine
                w[1:] = (1-n_t*stopa)*w[1:] + (n_t*y_i*x_i[1:])

                #Računanje biasa
                w[0] = w[0] + (y_i*x_i[0])
            else:
                w = (1-n_t*stopa)*w

            min_distance_positive, min_distance_negative, konvergira = self.provjeri_potporni(skup_treniranja, w) #Provjera na kojoj je strani hiperravni potporni vektor
        print("<------------------>\n")
        print("Broj iteracija do konvergiranja: ", k)
        
        self.prikazi(skup_treniranja, skup_testiranja,min_distance_negative, min_distance_positive, w,"Train Plot", "", "")
        
        return w, min_distance_negative, min_distance_positive

    def provjeri_potporni(self, skup_treniranja, w):
        """

        Here, we identify the support vectors for each weight vectort and see if it is equidistant from w

        """

        min_distance_positive = 999.0
        min_distance_negative = 999.0

        for i in skup_treniranja:
            x1 = i[1]
            x2 = i[2]
            y = i[3]
            try: #formula w/||w||
                d = abs(w[1]*x1 + w[2]*x2 + w[0])/ (math.sqrt(w[1]**2) + (math.sqrt(w[2]**2)))
                if y == -1:
                    if d<=min_distance_negative:
                        min_distance_negative = d
                else:
                    if d<=min_distance_positive:
                        min_distance_positive = d
            except: 
                pass

        if round(min_distance_positive,1) == round(min_distance_negative,1):
            return round(min_distance_positive)+0.6, round(min_distance_negative)+0.6, True
        else:
            return 1,1,False

    def prikazi(self, skup_treniranja, skup_testiranja, min_distance_negative, min_distance_positive, w, plot_type, naziv1, naziv2):
        """
        Jednostavna funkcija koja prelazi kroz podatke i iscrtava ih koristeći matplotlib biblioteku

        """

        if plot_type == "Train Plot":
            plt.scatter(skup_treniranja[:, 1], skup_treniranja[:, 2], c=skup_treniranja[:, 3],  edgecolor="black")
            ax = plt.gca()
            ax.set_facecolor('#ffeadb')
            ax.patch.set_alpha(0.5)
            xlim = ax.get_xlim()

            xx = np.linspace(xlim[0], xlim[1])
            yy = -(w[1]/w[2]) * xx - (w[0]/w[2])
            yy1 = min_distance_positive-(w[1]/w[2]) * xx - (w[0]/w[2])
            yy2 = -min_distance_negative-(w[1]/w[2]) * xx - (w[0]/w[2])
            plt.plot(xx, yy, c="r")
            plt.plot(xx, yy1, c="g", linestyle = "dashed")
            plt.plot(xx, yy2, c="g", linestyle = "dashed")
            plt.show()
        if plot_type == "Test Plot":
            plt.scatter(skup_treniranja[:, 1], skup_treniranja[:, 2], c=(skup_treniranja[:, 3]), edgecolor="black")
            plt.scatter(skup_testiranja[:, 1], skup_testiranja[:, 2], c='r', marker="D", edgecolor="black")
            ax = plt.gca()
            ax.set_facecolor('#ffeadb')
            ax.patch.set_alpha(0.5)
            xlim = ax.get_xlim()

            xx = np.linspace(xlim[0], xlim[1])
            yy = -(w[1]/w[2]) * xx - (w[0]/w[2])
            yy1 = min_distance_positive-(w[1]/w[2]) * xx - (w[0]/w[2])
            yy2 = -min_distance_negative-(w[1]/w[2]) * xx - (w[0]/w[2])
            plt.plot(xx, yy, c="r")
            plt.plot(xx, yy1, c="g", linestyle = "dashed")
            plt.plot(xx, yy2, c="g", linestyle = "dashed")
            
            predikcija_negativan = mpatches.Patch(color='darkblue', label=naziv1)
            predikcija_pozitivan = mpatches.Patch(color='y', label=naziv2)
            plt.legend(handles=[predikcija_negativan, predikcija_pozitivan])

            return plt


    def test(self, w, min_distance_negative, min_distance_positive, plot_type, naziv1, naziv2):
        """
        Funkcija za testiranje SVM Modela

        """
        plt = self.prikazi(self.skup_treniranja, self.skup_testiranja, min_distance_negative, min_distance_positive, w, plot_type, naziv1, naziv2)

        greske = 0
        for i in self.skup_testiranja:
            predikcija = np.dot(i[:-1], w)
            if predikcija<0 and i[-1] == 1:
                greske +=1
                plt.scatter(i[1], i[2], c='purple', edgecolor="white")
            elif predikcija>0 and i[-1] == -1:
                greske +=1
                plt.scatter(i[1], i[2], c='orange', edgecolor="white")
        print("\nUkupno tačaka trenirano: ", len(self.skup_treniranja))
        print("\nUkupno tačaka testirano: ", len(self.skup_testiranja))
        print("\nBroj pogrešnih tačaka: ", greske)
        print("\nPreciznost: ", ((len(self.skup_testiranja) - greske) / len(self.skup_testiranja)) * 100, "%")
        print("\n<------------------>")

        plt.show()
```

```python
%matplotlib inline
import numpy as np

setosa_data = [[a,b] for a, b in zip(setosa_x, setosa_y)]
virginica_data = [[a,b] for a, b in zip(vriginica_x, vriginica_y)] 
#Postavljanje podataka, -1 - setosa
#                        1 - virginica


#Generisanje skupa podataka za testiranje i obučavanje
svm = MetodaPotpornihVektora(setosa_data, virginica_data)
#Nalaženje vektora težine i potpornih vektora na 150000 iteracija
w, min_distance_negative, min_distance_positive = svm.train()
#Testiranje SVM Modela
svm.test(w, min_distance_negative, min_distance_positive, "Test Plot", "Setosa", "Virginica")
```

# Opis logističke regresije

<div style="text-align: justify">
Dok u metodi potpornih vektora algoritam pokušava da nađe "najbolju" marginu za klasifikaciju podataka, kod logističke regresije situacija nije takva. Logistička regresija ima drugačije granice sa različitim težinama koji su blizu optimuma. Kao i kod metode potpornih vektora, algoritam logističke regresije koristimo za nalaženje (predikciju) vjerovatnoće datog podatka (klase) i njegovog ishoda zavisno od treniranih podataka. Vidjeli smo da kod SVM algoritma za neki ulazni podatak A dobijamo rezultat B u obliku vrijednosti 1 ili 0, isti slučaj imamo i kod logističke regresije. Međutim, način na koji pronalazi to rješenje se vrši preko logističke (sigmoidne) funkcije da nađe vezu između ulaznih podataka. Sigmoidna funkcija je kriva oblika S na dvodimenzionalnom sistemu koja prima bilo koji input (ulaz) i stavlja između vrijednosti (0,1). Za prikaz klasifikacije na grafu koristimo granicu odluke, dok sigmoidna funkcija je ta funkcija koja će vršiti klasifikaciju.

Slika 7. Predstavljamo funkciju troška u pythonu:


Minimizacija troška se vrši ažuriranjem novih parametara koje ta funkcija prima koristeći funkciju gradijentnog spusta. Cilj gradijentnog spusta je da nađe optimalne parametre, u logističkoj regresiji, nalazeći minimum funkcije.


```python
import numpy as np

def funkcija_troska(h, y):
    return np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))

funkcija_troska(1/2, 1)
```

```python
#Sigmoidna funkcija (postavlja vrijednosti između 0 i 1)
import math
def sigmoid_funkcija(vrijednost):
    rezultat = 1/(1+math.e**(-vrijednost))
    return rezultat

print(sigmoid_funkcija(10), sigmoid_funkcija(1))
```

<div style="text-align: justify">      
Pozivanjem sigmoidne funkcije u algoritmu nastojimo dobiti predikciju da neka vrijednost A pripada klasi 1 ili 0 zavisno da li je predikcija sigmoidne funkcije >= 0.5 ili < 0. Pored sigmoidne funkcije, logistička regresija nastoji smanjiti funkciju troška pomoću gradijenta spusta. Tu postavljamo brzinu obučavanja, ukoliko je vrijednost prevelika možemo "prevazići" optimalnu vrijednost. Brzina obučavanja α odlučuje koliko brzo će doći do ažuriranja novih parametara, s toga ukoliko je α premala dolazi do potrebe većeg broja iteracija za konvergiranje do te optimalne vrijednosti. Hipoteza logističke regresije postavlja i vrši limitaciju vrijednosti funkcije troška između 0 i 1. Linearne funkcije će imati vrijednosti koje ne spadaju u ovu domenu i samim tim se ne mogu koristiti. <br/> Hipotezu u logističkoj regresiji predstavljamo kao:
</div>
    
<math>σ(Z) = σ(β₀ + β₁X)</math> <br/>
Primjer jedne funckije logističke regresije je:
y = e^(β₀ + β₁*x) / (1 + e^(β₀ + β₁*x))
gdje je β₀ bias, a β₁ koeficijent za ulazni podatak x. Ubacivanjem podataka u ovu funkciju dobit ćemo neku vrijednost iz koje možemo predpostaviti klasu za taj podatak.
<div style="text-align: justify">     
što je malo drugačije od formule za linearnu regresiju
<math>hΘ(x) = β₀ + β₁X</math>
Za vraćanje vrijednosti između 0 i 1 u logističkoj regresiji koristimo sigmoidnu funkciju nad datom hipotezom da dobijemo rezultat za klasifikaciju.
</div>

```python
import numpy as np
import time

#Postavljamo brzinu obučavanja kao i broj iteracija.

class LogistickaRegresija:
    def __init__(self, X, brzina_obucavanja=0.1, broj_iteracija=15000):
        self.brzina_u = brzina_obucavanja
        self.broj_it = broj_iteracija

        # m je broj podataka, dok je n broj osobina
        self.m, self.n = X.shape

    def train(self, X, y):
        # inicijalizacija tezine i biasa/slobodnog člana
        self.tezinski_vektor = [[0. for kolone in range(1)] for red in range(self.n)]
        self.bias = 0

        #treniranje kroz petlju za num_iteracija
        for it in range(self.broj_it + 1):
            #računanje hipoteze
            y_predict = sigmoid_funkcija(np.dot(X, self.tezinski_vektor) + self.bias)

            # cilj umanjivanja troska/cijene
            cijena = -1 / self.m * funkcija_troska(y_predict, y) #optimizacija

            # racunanje gradijent spusta (glavni cilj je minimiziranjem troska)
            #skalarni proizvod transponovane matrice X sa y_predict - y
            dw = 1 / self.m * np.dot(X.T, (y_predict - y))
            db = 1 / self.m * np.sum(y_predict - y)

            # gradient descent azuriraj korak
            self.tezinski_vektor -= self.brzina_u * dw
            self.bias -= self.brzina_u * db

            # nakon 2000-te iteracije ispisati trosak
            if it % 2000 == 0:
                print(f"Trošak iteracije {it} -> {cijena}")
                
        self.y_predict = y_predict

        return self.tezinski_vektor, self.bias

    def predikcija(self, X):
        y_predict = sigmoid_funkcija(np.dot(X, self.tezinski_vektor) + self.bias)
        y_predict_rezultat = y_predict > 0.5 #provjera da li je veće od 0.5, u tom slučaju pripada klasi 1

        return y_predict_rezultat
    
    def prikaz_podataka(self, dataset1, dataset2, X, y, y_predict, naziv1, naziv2):
        trening_set_velicina = len(y)-len(y_predict)
        ax = plt.gca()
        #Prikaz rezultata na grafu
        x_vrijednost = [np.array(dataset1).min(),np.array(dataset2).max()]
        y_vrijednost = -(x_vrijednost * w[0] + b) / w[1] 
        
        
        ax.title.set_text("Prikaz granice odluke logističke regresije i predikcija")
        ax.set_facecolor('#ffeadb')
        ax.patch.set_alpha(0.5)
        xlim = ax.get_xlim()

        xx = np.linspace(xlim[0], xlim[1])
        
        plt.scatter(dataset1[0],dataset1[1],marker='D',color='blue', edgecolor="black", label=naziv1)
        plt.scatter(dataset2[0],dataset2[1],marker='o',color='red', edgecolor="black", label=naziv2)

        for i in range(len(y_predict)):
            if (y[trening_set_velicina + i]!=y_predict[i]):
                if (y[i] == 0):
                    ax.scatter(X[trening_set_velicina + i][0],X[trening_set_velicina+i][1],marker='x',color='orange')
                else:
                    ax.scatter(X[trening_set_velicina+i][0],X[trening_set_velicina+i][1],marker='x',color='green')
            else:
                ax.scatter(X[trening_set_velicina+i][0],X[trening_set_velicina+i][1],marker='d',color='white', edgecolor="black")  
        ax.legend()  
        ax.plot(x_vrijednost,y_vrijednost, c="r")
        
```

Linearni i polinomijalni kernel zahtijevaju manje vremena sa manom što donose lošiju preciznost klasifikacije, dok rbf kernel radi suprotno. RBF koristi normalne krive oko skupa podataka koje spaja da dobije granicu odluke. Kerneli "mapiraju" date ulazne podatke u veću dimenziju u kojoj tada ti podaci postanu linearno separabilni.

#### Linear Kernel: K(X,Y)=X<sup>T</sup>Y <br/>
#### Polynomial Kernel: K(X,Y)=(γ⋅X<sup>T</sup>Y+r)<sup>d</sup>,γ>0 <br/>
#### Radial basis function (RBF) Kernel: K(X,Y)=exp(∥X−Y∥<sup>2</sup>/2σ<sup>2</sup>) koji se može zapisati u formi exp(−γ⋅∥X−Y∥<sup>2</sup>),γ>0 <br/>



```python
from sklearn import svm, datasets
from sklearn.model_selection import train_test_split
import time

iris = datasets.load_iris()
X = iris.data[:100, :2]
y = iris.target[:100]

def sklearnBibliotekeSVM(X, y, ravel=False):
    h = .02  # step size in the mesh
    x_train,x_test,y_train,y_test=train_test_split(X,y,train_size=0.8,test_size=0.2)
    C = 100.0  # SVM regularization parameter
    svc = svm.SVC(kernel='linear', C=C).fit(X, y)
    start = time.process_time()
    if ravel:
        rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(x_train, y_train.values.ravel())
        print(f'Vrijeme treniranja RBF Kernel SVM-a: {time.process_time() - start}s')
        start = time.process_time()
        poly_svc = svm.SVC(kernel='poly', degree=3, C=C).fit(x_train, y_train.values.ravel())
        print(f'Vrijeme treniranja Polinomijalnog Kernel SVM-a: {time.process_time() - start}s')
    else:
        rbf_svc = svm.SVC(kernel='rbf', gamma=0.7, C=C).fit(x_train, y_train)
        print(f'Vrijeme treniranja RBF Kernel SVM-a: {time.process_time() - start}s')
        start = time.process_time()
        poly_svc = svm.SVC(kernel='poly', degree=3, C=C).fit(x_train, y_train)
        print(f'Vrijeme treniranja Polinomijalnog Kernel SVM-a: {time.process_time() - start}s')

    titles=["Metoda potpornih vektora sa RBF Kernel funkcijom", "Metoda potpornih vektora sa Polinomijalnom Kernel funkcijom"]

    # mesh plot
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    plt.subplots_adjust(wspace=0.4, hspace=0.4)
    start = time.process_time()

    naziv= ["RBF", "Polinomijalnog"]
    print('\n')
    for i, clf in enumerate((rbf_svc, poly_svc)):  
        plt.subplot(1, 1 + i, i + 1)
        plt.subplots_adjust(wspace=0.4, hspace=0.4)
        start = time.process_time()
        Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
        print(f'Preciznost: {clf.score(x_test, y_test)}')
        print(f'Vrijeme predikcije {naziv[i]} Kernel SVM-a: {time.process_time() - start}s')
        Z = Z.reshape(xx.shape)
        plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)

        # plot treniranja
        plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm)
        plt.xlim(xx.min(), xx.max())
        plt.ylim(yy.min(), yy.max())
        plt.xticks(())
        plt.yticks(())
        plt.title(titles[i])

        plt.show()
        
sklearnBibliotekeSVM(X, y)    
```

Pošto su klase vidljivo različite bez velikih odstupanja, metode bi trebale imati visoku preciznost klasifikacije i na manjem broju treniranja. U ovom slučaju posmatramo 1000 ovakvih podataka što nam daje:
</div>

```python
#Klasifikacija - gaussian isotrpic blobs
from sklearn.datasets import make_blobs

np.random.seed(1)

center_box = (0, 50) # okolika u kojoj ce se nalaziti
standard_dev = 3 # standardna devijacija podataka (veće odstupanje)

X, y = make_blobs(n_samples=1000, centers=2, center_box=center_box, cluster_std=standard_dev)
y = y[:, np.newaxis]

print("Prvih 10 podataka dataset-a:\n", X[:10], "\nKlasa:\n", list(y[:10]))

blob_klasa_jedan = []
blob_klasa_nula = []
for index, klasa in enumerate(y):
    if (klasa == 0):
        blob_klasa_nula.append(X[index])
    else:
         blob_klasa_jedan.append(X[index])
            
plt.scatter([item[0] for item in blob_klasa_nula],[item[1] for item in blob_klasa_nula],marker='+',color='red', label='Klasa 1')
plt.scatter([item[0] for item in blob_klasa_jedan],[item[1] for item in blob_klasa_jedan],marker='o',color='blue', label="Klasa 2")
plt.legend()
```

Skup podataka seta "gaussian isotropic blobs" se sastoji od 2 liste X i y. Radi se o binarnoj klasifikaciji. ## Zaključak

<div style="text-align: justify">
SVM algoritam je dizajniran za generisanje kompleksnijih granica odluke. U ovim primjerima LS-SVM je koristio funkciju linearnog kernela što doprinosi linearnoj granici odluke. Pored linearnog kernela, postoje i kompleksnije funkcije kao RBF kernel koji dodaje dodatni parametar koji odlučuje zakrivljenje kod granice odluke. SVM je relativno nova metoda koja je bazirana na principu statičke teorije učenja. Ovakva metoda pokušava da obučava i generalizuje veoma dobro kod pravljenja modela sa datim setom podataka, dok se kod predikcije kod prijašnje neviđenih podataka dolazi do malog usporenja performanse što se i vidi u primjerima, gdje je logističa regresija imala puno bolju brzinu. Logistička regresija radi puno bolje na problemima sa manjom dimenzijom, kao u slučaju ovog projekta. U većini novih podataka, logistička regresija je imala puno bolju brzinu i performansu. U nekim grafovima vidjeli smo da "anomalije" su imale puno veći uticaj kod logističke regresije nego kod SVM mašine jer funkcija troška puno brže divergira neko kod SVM-a. Kod podataka koji su bili linearno separabilni (kao Iris ili Gaussian Blob) logistička regresija je puno bolje rezultate davala nego SVM mašina, u suprotnom SVM sa drugim kernelom bi imao tačnija rješenja ukoliko podaci ne bi imali linearnu separablinost. 
<br/><br/>
Glavna činjenica kod SVM algoritma je da ukoliko za predikciju imamo skup podataka koji nije linearno separabilan ili ima očiglednih pogrešnih pozitivnih/negativnih predikcija, korištenjem kompleksnijeg kernala u algoritmu dobit ćemo i bolje rezultate. Korištenjem takvih funkcija kernela zahtjeva više memorije kao i duže vrijeme predikcije i obuke što može biti izuzetno neefikasno ukoliko imamo ogroman skup podataka. Samim tim ukoliko je takav skup podataka moguče linearno klasificirati kao početni podaci u projektu, tada je uvijek bolje koristiti neki od prvobitnih algoritama i metoda klasifikacije. Logistička regresija i SVM su veoma pouzdani algoritmi, i kroz neke primjere binarne klasifikacije vidimo kako sa velikom preciznošću daju rezultate koje ne bi očekivali za podatke koje vidimo na grafu. Zavisno od kakvog dataset-a se radi logistička regresija i linearni SVM nemaju velike promijene između sebe, dok kod kompleksnijih podataka uvijek je bolje preći na druge kernele ili tražiti daljnje algoritme za klasifikaciju.
