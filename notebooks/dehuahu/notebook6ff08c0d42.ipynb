**CREO UN NUEVO DF=datos CON LAS MEAN Y STD DONDE LOS INDICES/FILAS SON LAS COLUMNAS CUANTITATIVAS. DESPUES INTRODUZCO 4 COLUMNAS: MEDIA+STD, MEDIA-STD, MEDIA+2*STD, MEDIA-2*STD. DESPUES CALCULO LOS REGISTROS DE CADA COLUMNA QUE ESTEN EN LOS RANGOS DE CADA FILA DE datos (index=columnas). FINALMENTE AÃ‘ADO LOS PORCENTAJES PARA CADA COLUMNA QUE ESTEN DENTRO DE LOS RANGOS**

```python
datos=pd.DataFrame({"promedio":lista_promedio, "desviacion":lista_desviacion})
datos.head()
```

```python
lista_una_mas=[]
lista_una_menos=[]
lista_dos_mas=[]
lista_dos_menos=[]
for i in range(len(datos)):
    uno_mas=datos.iloc[i,0]+datos.iloc[i,1]
    uno_menos=datos.iloc[i,0]-datos.iloc[i,1]
    lista_una_mas.append(uno_mas)
    lista_una_menos.append(uno_menos)
    dos_mas=datos.iloc[i,0]+(2*(datos.iloc[i,1])) 
    dos_menos=datos.iloc[i,0]-(2*(datos.iloc[i,1]))
    lista_dos_mas.append(dos_mas)
    lista_dos_menos.append(dos_menos)
```

```python
datos["mas1"]=lista_una_mas
datos["menos1"]=lista_una_menos
datos["mas2"]=lista_dos_mas
datos["menos2"]=lista_dos_menos
```

```python
porcentaje_1=[]
porcentaje_2=[]
for i in range(len(cuantitativos)):
    cantidad_1=df[df_copia.iloc[:,i].between(datos.iloc[i, 3],datos.iloc[i, 2])].shape
    cantidad_2=df[df_copia.iloc[:,i].between(datos.iloc[i, 5],datos.iloc[i, 4])].shape
    porcentaje_1.append(cantidad_1[0]/1782*100)
    porcentaje_2.append(cantidad_2[0]/1782*100)
```
