```python
import numpy as np
import pandas as pd
import plotly.express as px
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
covid_vaccination_filepath = "../input/covid-world-vaccination-progress/country_vaccinations.csv"

covid_vaccination_data = pd.read_csv(covid_vaccination_filepath)
```

```python
covid_vaccination_data
```

Which contry has vacciated the most people so far?

```python
covid_vaccinations = covid_vaccination_data.dropna()
covid_vaccinations

covid_vaccinations = covid_vaccinations.sort_values(by=['total_vaccinations'])
covid_vaccinations = covid_vaccinations.tail(200)
```

```python
plt.figure(figsize=(8,8))

plt.title("Top 10 Vaccinated Countries")

sns.barplot(x=covid_vaccinations["total_vaccinations"], y=covid_vaccinations["country"], palette="Blues_d")

plt.xlabel("Total Vaccinations")
plt.ylabel("Country")
```

Total Vaccinations by Country

How many people is the USA vaccinating each day?

```python
covid_vaccinations_USA = pd.read_csv(covid_vaccination_filepath, index_col="iso_code")

covid_vaccinations_USA = covid_vaccinations_USA.loc["USA"]
covid_vaccinations_USA = covid_vaccinations_USA.fillna(0)
```

```python
plt.figure(figsize=(12,20))

plt.title("United States Daily Covid Vaccinations")

sns.barplot(x=covid_vaccinations_USA["total_vaccinations"], y=covid_vaccinations_USA["date"], palette="Blues_d")

plt.xlabel("Daily Vaccinations")
plt.ylabel("Date")
```

How many people is Brazil vaccinating each day?

```python
covid_vaccinations_Brazil = pd.read_csv(covid_vaccination_filepath, index_col="iso_code")

covid_vaccinations_Brazil = covid_vaccinations_Brazil.loc["BRA"]
covid_vaccinations_Brazil = covid_vaccinations_Brazil.fillna(0)
```

```python
plt.figure(figsize=(16,12))

plt.title("Brazil Daily Covid Vaccinations")

sns.barplot(y=covid_vaccinations_Brazil["daily_vaccinations"], x=covid_vaccinations_Brazil["date"], palette="Blues_d")

plt.xticks(rotation=45)
plt.xlabel("Daily Vaccinations")
plt.ylabel("Date")
```

Which country has vaccinated the highest percentage of their population

```python
covid_vaccinations_hundred = pd.read_csv(covid_vaccination_filepath, index_col="date")
covid_vaccinations_hundred = covid_vaccinations_hundred.loc["2021-03-03"]
covid_vaccinations_hundred = covid_vaccinations_hundred.set_index("people_vaccinated_per_hundred")
covid_vaccinations_hundred = covid_vaccinations_hundred.dropna()
covid_vaccinations_hundred = covid_vaccinations_hundred.sort_index()
covid_vaccinations_hundred

#covid_vaccinations_hundred = covid_vaccinations_hundred.sort_values(by=['total_vaccinations_per_hundred'])
```

```python
plt.figure(figsize=(12,8))

plt.title("Vaccinations Per Hundred")

sns.barplot(y=covid_vaccinations_hundred["total_vaccinations_per_hundred"], x=covid_vaccinations_hundred["country"], palette="Blues_d")

plt.xticks(rotation=90)
plt.ylabel("Total Vaccinations Per Hundred")
plt.xlabel("Country")
```

The above graph gives a rough representation of the percentage of the population that these countries have vaccinated. Many of the countries scoring higher began vacciating their populations first, or have much smaller populations than other countries
