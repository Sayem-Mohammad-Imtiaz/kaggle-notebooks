# hypothesis test (가설검정)

```python
plt.figure(figsize=(20,15))
plt.subplot(3,2,1)

sns.boxplot(x='gender',y="math score", data=data)
plt.title('math score(boxplot)')
plt.subplot(3,2,2)
sns.violinplot(x='gender',y="math score", data=data)
plt.title('math score(violinplot)')


plt.subplot(3,2,3)
sns.boxplot(x='gender',y="writing score", data=data)
plt.title('writing score(boxplot)')
plt.subplot(3,2,4)
sns.violinplot(x='gender',y="writing score", data=data)
plt.title('writing score(violinplot)')


plt.subplot(3,2,5)
sns.boxplot(x='gender',y="reading score", data=data)
plt.title('reading score(boxplot)')

plt.subplot(3,2,6)
sns.violinplot(x='gender',y="reading score", data=data)
plt.title('reading score(violinplot)')
```

```python
data.groupby('gender').mean().style.background_gradient(cmap = "OrRd")
```

```python
data_m_math = data[data['gender']=='male']
data_f_math = data[data['gender']=='female']
```

## 1, Hypothesis test of math score by gender
## 1, 성별에 따른 수학성적 가설검정

Ho = There is no difference in math scores between genders

H1 = Differences in math scores between genders exist.

Ho = 성별간의 수학점수 차이는 없다

H1 = 성별간의 수학점수 차이는 존재 한다.

```python
scipy.stats.ttest_ind(data_m_math['math score'], data_f_math['math score'], equal_var=False)
```

Conclusion -> The p value is 8.42083810e-08(p<0.05), so there is a difference in math scores between genders. Therefore, men are better at math than women.

결론 -> p값이 8.42083810e-08(p<0.05) 이므로 성별간의 수학 점수 차이가 존재한다. 따라서 남자가 여자보다 수학 성적이 우수하다고 볼 수 있다. 

## 2, Hypothesis test of reading score by gender
## 2, 성별에 따른 읽기성적 가설검정

Ho = There is no difference in listening scores between genders

H1 = Differences in listening scores between genders exist.

Ho = 성별간의 듣기점수 차이는 없다

H1 = 성별간의 듣기점수 차이는 존재 한다.

```python
scipy.stats.ttest_ind(data_m_math['reading score'], data_f_math['reading score'], equal_var=False)
```

Conclusion -> P is 4.37629e-15(p<0.05), so there is a difference in reading score between genders. Therefore, women can be considered to have better listening performance than men.

결론 -> p값이 4.37629e-15(p<0.05) 이므로 성별간의 듣기 성적차이가 존재한다. 따라서 여자가 남자보다 듣기성적이 우수하다고 볼 수 있다.

## 3, Hypothesis test of writing score by gender
## 3, 성별에 따른 쓰기성적 가설검정

Ho = There is no difference in writing scores between genders.

H1 = Differences in writing scores between genders exist.

Ho = 성별간의 쓰기점수 차이는 없다

H1 = 성별간의 쓰기점수 차이는 존재 한다.

```python
scipy.stats.ttest_ind(data_m_math['writing score'], data_f_math['writing score'], equal_var=False)
```

Conclusion -> Because the p value is 1.711809371e-22(p0.00.05), there is a difference in written performance between genders. Therefore, women are better at writing than men.

결론 -> p값이 1.711809371e-22(p<0.05) 이므로 성별간의 쓰기성적 차이가 존재한다. 따라서 여자가 남자보다 쓰기성적이 우수하다고 볼 수 있다.
