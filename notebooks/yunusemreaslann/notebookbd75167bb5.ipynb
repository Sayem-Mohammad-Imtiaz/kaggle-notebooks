**Preparation**

```python
df = pd.read_csv("/kaggle/input/bitcoin-historical-data/bitstampUSD_1-min_data_2012-01-01_to_2021-03-31.csv")
df.info()
```

```python
df.head()
```

```python

df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='s')

df["time"] = df['Timestamp'].dt.date # better display
df.head()
```

```python
df.corr()
```

**Weighted Price Over The Years**

```python
df.plot(kind='line', x='time', y='Weighted_Price',alpha = 0.4,color = 'blue',grid= True, xlabel = "Time", ylabel = "Weighted Price", title = "Weighted Price Over The Years", label = "weighted price") # weighted price over the years
plt.legend()
plt.show()
```

**Correlation Between Open And Close Prices**

```python
df.plot(kind='line', x='Open', y='Close',alpha = 0.4,color = 'blue',grid= True, xlabel = "Open Prices", ylabel = "Close Prices", title = "Correlation Between Open And Close", label = "corrolation line")
plt.legend()
plt.show()
```
