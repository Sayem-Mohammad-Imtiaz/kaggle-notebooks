**Pivoting Data Frames**

```python
dic = {"name":["Tip1","Tip1","Tip2","Tip2"], "age": [13,23,14,12],"country":["turkey","usa","uk","brazil"],"class":[1,3,4,6]}
data2 = pd.DataFrame(dic)
data2
```

```python
data2.pivot(index = "name", columns = "country",values = "class")
```

**Stacking And Unstacking DataFrame**

```python
data3 = data2.set_index(["name","country"])
data3
```

```python
data3.unstack(level = 0)
```

```python
data3.unstack(level = 1)
```

```python
data4 = data3.swaplevel(0,1)
data4
```

```python
pd.melt(data2, id_vars="name", value_vars=["age","country"])
```
