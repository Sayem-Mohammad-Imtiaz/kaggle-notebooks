```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
df=pd.read_csv('/kaggle/input/breast-cancer-wisconsin-data/data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.columns
```

```python
df.drop(['Unnamed: 32','id'],axis=1,inplace=True)
```

```python
Y=df['diagnosis']
X=df.drop(['diagnosis'],axis=1)
```

```python
print(X.shape)
print(Y.shape)
```

```python
X_cols=X.columns
```

```python
sns.countplot(Y)
```

```python

Y=Y.map({'B':0,'M':1})
```

```python
Y.value_counts()
```

```python
from sklearn.preprocessing import RobustScaler
sc=RobustScaler()
transformed=sc.fit_transform(X)
transformed
```

```python
X_transformed=pd.DataFrame(transformed,columns=X_cols)
```

```python
X_transformed
```

```python
sns.distplot(X_transformed)
```

```python
model=Sequential()
model.add(Dense(32,input_dim=30,activation='relu',kernel_initializer='random_normal'))
model.add(Dense(64,activation='relu',kernel_initializer='random_normal'))
model.add(Dense(128,activation='relu',kernel_initializer='random_normal'))
model.add(Dropout(0.3))

model.add(Dense(256,activation='relu',kernel_initializer='random_normal'))
model.add(Dense(512,activation='relu',kernel_initializer='random_normal'))
model.add(Dense(1024,activation='relu',kernel_initializer='random_normal'))
model.add(Dropout(0.3))

model.add(Dense(128,activation='relu',kernel_initializer='random_normal'))
model.add(Dense(256,activation='relu',kernel_initializer='random_normal'))
model.add(Dense(512,activation='relu',kernel_initializer='random_normal'))
model.add(Dropout(0.3))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='SGD',loss='binary_crossentropy',metrics=['binary_accuracy'])
```

```python
model.summary()
```

```python
model_result=model.fit(X_transformed,Y,batch_size=50,epochs=200,validation_split=0.2,verbose=1,shuffle=True)
```
