```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load data

```python
df = pd.read_csv('../input/bike-sharing-dataset/day.csv')
df.head()
```

```python
# Let's see the correlation between different variables using heatmap
fig, ax = plt.subplots(figsize=(20,20))
sns.heatmap(df.corr(), annot = True, ax=ax)
plt.show()
```

# Feature Engineering

```python
#feature extraction
features = ["season","yr","mnth","holiday","weekday","workingday","weathersit","temp","hum","windspeed", "casual", "registered"]
```

```python
X = df[features]
y = df["cnt"]
```

# Train Model

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size = 0.2)
```

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
```

```python
pred = rf.predict(X_val)
```

```python
rf.score(X_val, y_val)
```

# Predictive performance

```python
dataFramePredict = pd.DataFrame(pred , index = y_val.index)
```

```python
dataFramePredict
```

```python
idx = np.sort(y_val.index)
```

```python
idx
```

```python
y_true = y_val[idx]
y_pred = dataFramePredict[0][idx]
```

```python
y_true, y_pred
```

```python
plt.figure(figsize = (16, 7))
plt.plot(idx, y_true, label = "True")
plt.plot(idx, y_pred, label = "Predict")
plt.legend()
plt.show()
```
