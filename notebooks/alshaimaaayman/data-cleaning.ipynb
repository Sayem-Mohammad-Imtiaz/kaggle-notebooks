```python
# drop rows with null values 
df3 = df2.dropna()
df3.isnull().sum()
```

```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
# Get ranges from total_sqft column using above function
df3[~df3['total_sqft'].apply(is_float)].head()
```

```python
#convert 'total_sqft' values to float numbers and save it in anther file
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try :
        return float(x)
    except:
        return None
```

```python
df4 = df3.copy()
```

```python
df4['total_sqft'] = df4['total_sqft'].apply(convert_sqft_to_num)
```
