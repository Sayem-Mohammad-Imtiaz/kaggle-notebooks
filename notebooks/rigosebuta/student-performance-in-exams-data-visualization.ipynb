# Data Visualization

```python
# gender distribution
fig1 = px.pie(stud_data, names = 'gender', title = "Overall gender distribution")
fig1.show()
```

 => more female students than male students

```python
stud_data_race_gender = stud_data.groupby(["race/ethnicity", 'gender']).count().reset_index()
stud_data_race_gender = stud_data_race_gender.rename(
    columns = {'parental level of education': 'counter'} )
stud_data_race_gender = stud_data_race_gender.drop(['lunch', 'test preparation course', 'math score',
                            'reading score', 'writing score'], axis = 1)
#stud_data_race_gender.head(10)
fig2 = px.bar(stud_data_race_gender, x = 'race/ethnicity', y = "counter", color = 'gender',
             barmode = 'group', title = "Gender distribution between different ethnicities")
fig2.show()
```

 => more males in group A, but more females in group B and group C

```python
stud_data_education = stud_data.groupby(['parental level of education', 'gender']).count().reset_index()
stud_data_education = stud_data_education.rename(columns = {'race/ethnicity': 'counter'})
stud_data_education = stud_data_education.drop(['lunch','test preparation course', 'math score',
                            'reading score', 'writing score'], axis = 1)
stud_data_education.head(30)

education_ascending = {'some high school': 0, 'high school': 1, 'some college': 2, 'college': 3,
               "'associate's degree'":4, "'bachelor's degree'": 5, "'master's degree'": 6}
               
stud_data_education = stud_data_education.sort_values(by=['parental level of education'],
                                                      key=lambda x: x.map(education_ascending))

fig3 = px.bar(stud_data_education, x = 'parental level of education', y = "counter", color = 'gender',
             barmode = 'group', title = "Gender distribution between different education levels")
fig3.show()
```

=> females are more and more overrepresented as the education level increases

```python
stud_data_race_educ = stud_data.groupby(['parental level of education',
                                         'race/ethnicity']).count().reset_index()
stud_data_race_educ = stud_data_race_educ.rename(columns = {'gender': 'counter'})
stud_data_race_educ = stud_data_race_educ.drop(['lunch','test preparation course', 'math score',
                                                'reading score', 'writing score'], axis = 1)
#stud_data_race_educ.head(10)
education_ascending = {'some high school': 0, 'high school': 1, 'some college': 2, 'college': 3,
               "'associate's degree'":4, "'bachelor's degree'": 5, "'master's degree'": 6}
               
stud_data_race_educ = stud_data_race_educ.sort_values(by=['parental level of education'],
                                                      key=lambda x: x.map(education_ascending))
fig4 = px.bar(stud_data_race_educ,
              x = 'parental level of education', y = "counter",color = 'race/ethnicity',
             barmode = 'group', title = "Education distribution between different ethnicities")
fig4.show()
```

=> pupils of Group D tend to do more higher education than other pupils of Groups
=> pupils of Group tend to do less higher education than other pupils of Groups

```python
stud_data_gender_lunch = stud_data.groupby(['gender', 'lunch']).count().reset_index()
stud_data_gender_lunch = stud_data_gender_lunch.rename(columns = {'race/ethnicity': 'counter'})
stud_data_gender_lunch = stud_data_gender_lunch.drop(['parental level of education',
                                                'test preparation course', 'math score',
                                                'reading score', 'writing score'], axis = 1)
#stud_data_gender_lunch.head(10)
fig5 = px.bar(stud_data_gender_lunch,
              x = 'gender', y = "counter",color = 'lunch',
             barmode = 'group', title = "Lunch properties between genders")
fig5.show()
```

```python
stud_data_educ_lunch = stud_data.groupby(['parental level of education',
                                            'lunch']).count().reset_index()
stud_data_educ_lunch = stud_data_educ_lunch.rename(columns = {'race/ethnicity': 'counter'})
stud_data_educ_lunch = stud_data_educ_lunch.drop(['gender', 'test preparation course',
                                                      'math score', 'reading score',
                                                      'writing score'], axis = 1)
#stud_data_educ_lunch.head(10)

education_ascending = {'some high school': 0, 'high school': 1, 'some college': 2, 'college': 3,
               "'associate's degree'":4, "'bachelor's degree'": 5, "'master's degree'": 6}
               
stud_data_educ_lunch = stud_data_educ_lunch.sort_values(by=['parental level of education'],
                                                      key=lambda x: x.map(education_ascending))
fig6 = px.bar(stud_data_educ_lunch,
              x = 'parental level of education', y = "counter",color = 'lunch',
             barmode = 'group', title = "Lunch properties between education levels")
fig6.show()
```

=> for bachelorand master students the percentage of students who get free/reduced lunch is increased compared to the other education levels.

```python
stud_data_race_lunch = stud_data.groupby(['race/ethnicity',
                                            'lunch']).count().reset_index()
stud_data_race_lunch = stud_data_race_lunch.rename(columns = {'parental level of education': 'counter'})
stud_data_race_lunch = stud_data_race_lunch.drop(['gender', 'test preparation course', 'math score',
                                                  'reading score', 'writing score'], axis = 1)
stud_data_race_lunch.head(10)
fig7 = px.bar(stud_data_race_lunch,
              x = 'race/ethnicity', y = "counter",color = 'lunch',
             barmode = 'group', title = "Lunch properties between ethnicities")
fig7.show()
```

=> percentage of students who get free/reduced lunch decreases from A to E.

```python
stud_data_gender_test = stud_data.groupby(['gender',
                                          'test preparation course']
                                        ).count().reset_index()
stud_data_gender_test = stud_data_gender_test.rename(
    columns = {'parental level of education': 'counter'})
stud_data_gender_test = stud_data_gender_test.drop(['lunch','race/ethnicity', 
                                                    'math score',
                                                  'reading score', 'writing score'],
                                                   axis = 1)
stud_data_gender_test.head(10)
fig8 = px.bar(stud_data_gender_test,
              x = 'gender', y = "counter",color = 'test preparation course',
              barmode = 'group', title = "Participants of test courses by gender")
fig8.show()
```

```python
stud_data_race_test = stud_data.groupby(['race/ethnicity',
                                          'test preparation course']
                                        ).count().reset_index()
stud_data_race_test = stud_data_race_test.rename(
    columns = {'parental level of education': 'counter'})
stud_data_race_test = stud_data_race_test.drop(['lunch', 'gender', 
                                                    'math score',
                                                  'reading score', 'writing score'],
                                                   axis = 1)
stud_data_race_test.head(10)
fig9 = px.bar(stud_data_race_test,
              x = 'race/ethnicity', y = "counter",color = 'test preparation course',
              barmode = 'group', title = "Participants of test courses by race")
fig9.show()
```

=> group D has a very low percentage of students who did test preparation
=> group A and group E have a high percentage of students who completed test preparation

```python
stud_data_lunch_test = stud_data.groupby(['lunch',
                                          'test preparation course']
                                        ).count().reset_index()
stud_data_lunch_test = stud_data_lunch_test.rename(
    columns = {'parental level of education': 'counter'})
stud_data_lunch_test = stud_data_lunch_test.drop(['race/ethnicity', 'gender',
                                                    'math score',
                                                  'reading score', 'writing score'],
                                                   axis = 1)
stud_data_lunch_test.head(10)
fig10 = px.bar(stud_data_lunch_test,
              x = 'lunch', y = "counter",color = 'test preparation course',
              barmode = 'group', title = "Participants of test courses by race")
fig10.show()
```

=> students who get free/reduced lunnch are more likely to complete test preparation

```python
score_educ = stud_data.groupby('parental level of education').mean().reset_index()

education_ascending = {'some high school': 0, 'high school': 1, 'some college': 2, 'college': 3,
               "'associate's degree'":4, "'bachelor's degree'": 5, "'master's degree'": 6}
               
score_educ = score_educ.sort_values(by=['parental level of education'],
                                                      key=lambda x: x.map(education_ascending))
fig13 = px.bar(score_educ, x = "parental level of education",
               y = ["math score", "reading score", "writing score"], barmode = "group")
fig13.show()
```

=> reading, writing and writing scores increase with higher levels of education
