***Daily percentage of the vaccinated population per Million***  
  
I selected this column because I would like to know that " which country was the fastest to conduct large scale vaccination activities ? " This kind of data show us how strong the leadership of the country's government is. Accoridng the map, North America and EU countries started large scale vaccinations very early. This is probably because they own the pharmaceutical company that developed the vaccine. I was surpised at the speed of Chile and Mongolia but I don't why they were so fast.  
The data is represented by the intensity of the red color. This is because I thought that a single color would better to represent the frequency of the data. And the range of the color is 0 to 10k because ten thousant may be a maximum value of daily vaccinations.

```python
df1 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df1.index = pd.to_datetime(df1['date'])
df1 = df1.sort_index()
#print(df1.columns)
#print(df1)
fig1 = px.choropleth(df1, locations='country',locationmode='country names',color='daily_vaccinations_per_million',hover_name='country',animation_frame='date',color_continuous_scale='Reds',range_color=(0.0,10000.0))
fig1.update_layout(title_text='Daily percentage of the vaccinated population per Million',title_x=0.45)
fig1.show()

pl.offline.plot(fig1,filename='vaccinations.html')
```

***"Summary"***  
An earthquake of magnitude 6.9 occurred on May 1.
Previous earthquakes occurred on February 13 and March 20, both with a magnitude of 7.1. These earthquakes are believed to be caused by the 9.0 magnitude Tohoku earthquake that occurred in 2011. It is astonishing that a single earthquake has changed the seismic landscape of northern Honshu and the Japan Trench.  
  
There have been far more earthquakes in the decade after the 2011 earthquake than in the decade before the 2011 earthquake.  Therefore the earthquake incidence rate is much higher than the pre-2011 level.
According to the data, in the 90 years prior to the Tohoku earthquake, the incidence of earthquakes of magnitude 6.8 or greater was 0.58 per year.In the decade after the Tohoku earthquake, the rate was 2.04 per year (3.5 times higher), and in the short period after February 13, the rate was 13.5 per year (another 6.5 times higher).
These increases in earthquake rates are judged to be due to the "Coulomb stress change" caused by the magnitude 9.0 earthquake, which brought the surrounding faults closer to rupture.  
  
The three magnitude 7 earthquakes are all aftershocks of the magnitude 9.0 Tohoku earthquake of 2011. Those aftershocks are not over yet, and there is a possibility that more major earthquakes will occur in the future. However, it will not be at an unprecedented speed like the last 80 days.
