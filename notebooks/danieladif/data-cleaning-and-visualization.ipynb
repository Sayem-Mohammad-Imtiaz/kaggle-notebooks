```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
filepath = '../input/heart-disease-uci/heart.csv'
df = pd.read_csv(filepath)
```

```python
#Check the dataset
df.head()
```

```python
#retrieve all columns in dataset and number of entries in the dataset
print(df.columns)
print(df.shape)
```

```python
#There are 303 rows in the dataframe, and 14 columns.
#We will see any missing value from the datasets.
df.isnull().sum()
```

```python
#we consider 1 is having heart disease and 0 is not having heart disease
df.target.value_counts()
```

```python
#Checking the distribution of age
sns.distplot(df['age'])
```

```python
#From the graph above we can see that the age distribution is normaly distributed.
#We will checking the correlation between resting blood preassure (trestbps), with serum cholesterol (chol)
#with respect to target column.
sns.lmplot(x='trestbps', y='chol', hue='target', data = df)
```

```python
#boxplot Plot of sex and age
sns.boxplot(df['sex'], df['age'])
```

```python
#comparing the cholesterol level among oldpeak
plt.figure(figsize=(12,12))
sns.barplot(df['oldpeak'],df['chol'])
```

```python
#Combingin two distribution graph
sns.jointplot('trestbps', 'chol', df, kind='kde')
```

```python
#checking the correlation of each columns
plt.figure(figsize=(12,12))
sns.heatmap(df.corr(), annot=True)
```
