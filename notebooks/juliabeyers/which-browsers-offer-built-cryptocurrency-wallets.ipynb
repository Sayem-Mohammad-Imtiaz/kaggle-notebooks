Independently of your investment’s size, you need a safe and secure cryptocurrency wallet. There are many [crypto wallets](http://wlxwallet.com/) nowadays, and they come in several forms — software wallets, hardware wallets, and paper wallets. Some are made exclusively for desktop computers, others are designed for smartphones. One type of crypto wallet proved to be far more convenient than others – browser addons. 


Namely, some crypto users find it useful to have one-click-away access to their funds. When they’re browsing something, they can reach out to the wallet icon and make instant transactions.

After looking into this, we’ve encountered a rather interesting phenomenon – browsers offering built-in cryptocurrency wallets. So far, only one browser managed to pull this off successfully, and that’s Opera.

Let’s take a look at the wallet offered by Opera, and what are your options if you use other browsers.


Opera’s built-in Crypto Wallet
Team Opera might not like the fact that their favorite browser started losing the race to other solutions for viewing web pages. Browsers such as Google Chrome and Mozilla Firefox have been regarded as the most popular options, especially for desktop users. Nevertheless, the team at opera didn’t give up, so they decided to attract a bunch of new tech-savvy audiences from the fintech sector by introducing a built-in cryptocurrency wallet.

According to the official press release, Opera began working on its project dubbed “Reborn 3”, which was later revealed to be Opera 60. It was officially released on April 9, 2019, and it included a crypto wallet, web 3.0 support, as well as a VPN and ad blocker. In fact, making both web 3.0 support and a wallet available, Opera became the first browser of its kind, making a significant step forward when it comes to the way we browse (or should we say use) the Internet.

Right now, there are three separate Opera wallets – Ethereum, Bitcoin, and Tron. As stated in this Coindesk report, there are more than 170,000 active wallet users on Opera, with the Ethereum wallet being the most popular option among the fans.

Apart from Opera, there are currently no other browsers that went so far and created built-in wallet extensions. Nevertheless, you can install these extensions to the browser you prefer, especially Chrome and Firefox, so let’s take a look at them.




MetaMask
Being widely regarded as the most popular crypto wallet browser extension, MetaMask has been able to store all kinds of cryptocurrencies, but with one significant limit – these are mostly ERC20 tokens.

Namely, apart from being a wallet, MetaMask is also your gateway to the Ethereum blockchain, meaning it only focuses on ETH and other related coins that are parts of projects built on the Ethereum platform. Meta mask can be installed as an add-on to Chrome, Firefox, and Brave.

Some of the good stuff that it comes with are open-source, built-in exchange, user-friendly interface, and excellent customer support. Still, the main disadvantage is that you cannot store Bitcoin or any other non-Ethereum cryptocurrency.


Jaxx
Jaxx is another popular option, and what makes it better than MetaMask is that it accepts other cryptos apart from Ethereum. On the other hand, it’s only available as a browser extension for Google Chrome, meaning you’ll have to seek alternatives if you use Firefox or any other browser.

Jaxx wasn’t originally built to be an extension. It also exists as a stand-alone platform for Windows, Mac, and Linux. You can even download it to your Android and iOS mobile devices. At the moment, you can store 90+ coins using this popular option.

Most importantly, Jaxx is completely free to use, meaning you’re one click away from installing it to your Chrome browser.

But, there’s something you should know:

However popular, Jaxx still has a huge stain on its name, as there was a huge security breach where a wallet user reportedly lost $400,000 when their account was hacked.


Exodus Crypto Wallet
Exodus is another popular option. It’s built as a stand-alone wallet but can be installed to Mozilla Firefox as an extension. It’s very easy to use and has professional customer support that works round the clock. Moreover, it supports a wide range of cryptocurrencies and comes with an intuitive user interface.

Still, the wallet lacks some property security protocols that are widely regarded as useful and used by many software wallets. On top of that, Exodus comes with fees that cannot really be described as competitive in the world of crypto wallets.

Conclusion: Is a Browser Extension the Best Option?
However useful and convenient, built-in browser wallets and wallet extensions don’t offer anything better compared to stand-alone wallets. In other words, the only thing that makes them “better” is the fact that they save you a bit of time, as you don’t always have to open the wallet when you want to make a transaction – you can do it in your browser. But we’re talking about a couple of additional seconds, which is something that you definitely want to spare if the product is good and the security level is high.

Still, you have to ask yourself — is that really necessary? The offer of wallets that aren’t available as browser extensions is much higher compared to the ones that are. Therefore, you’ll likely be able to find a wallet that suits you by taking all possible wallets into account than just focusing on the browser extensions.

The only option that’s really worth it is the one that comes with Opera since you’ll also gain access to web 3.0. Otherwise, make sure to read more about other wallers and find out which one you like.
