# Machine Learning

I would like to try some machine learning model on this data. We might find some interesting insigts from this dataset and try to implement it on real life.

We try different type of ML and then we choose which one is suitable with our needs.

**Naive Beyes model for Transfer fee and age**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

```python
X = data0001.iloc[:,[2]].values
Y = data0001.iloc[:,9].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state= 0)
```

```python
from sklearn.preprocessing import StandardScaler
scX = StandardScaler()
x_train = scX.fit_transform(x_train)
x_test = scX.fit_transform(x_test)
```

```python
classifier = GaussianNB()
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
cm = confusion_matrix(y_test, y_pred)

total = 0
for i in range(9):
 for j in range(9):
  total = total + cm[i,j]
accuracy = cm[2] + cm[9]
print('accuracy score: ', str(accuracy/total))
```

Maximum accuracy score for this model only 0.25 and it is not suitable for this dataset. We should try other model after this.

**Linear Regression Model**

```python
X = data0001.iloc[:, [2]].values
Y = data0001.iloc[:, 9].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state= 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)

plt.scatter(y_test, y_pred)
plt.title('Scatter plot of Transfer fee prediction')
plt.xlabel('Y test')
plt.ylabel('Y Prediction')
plt.show()
```

Based on current data we cannot predict transfer fee based on age. During season 2000-2001, most of the clubs doesnt care about age. The most importantly is capability of the player itself no matter young or old that player

**Conclusion**

On season 2000-2001 we cannot implement any type of machine learning model to predict transfer fee based on age because most of the club on this season doesnt care about age.

Club only thinks about capability of player no matter how age is he.
