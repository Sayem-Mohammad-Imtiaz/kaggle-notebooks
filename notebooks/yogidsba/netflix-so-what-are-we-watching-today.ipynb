### What is IMDb ratings? 
IMDb registered users can cast a vote (from 1 to 10) on every released title in the database. Users can update their votes as often as they'd like, but any new vote on the same title will overwrite the previous one, so it is one vote per title per user.
They take all the individual ratings cast by IMDb registered users and use them to calculate a single rating. IMDb don't use the arithmetic mean (i.e. the sum of all votes divided by the number of votes), although they do display the mean and average votes on the votes breakdown page; instead the rating displayed on a title's page is a weighted average. 
The IMDb weighted average does not change upon receipt of each new vote, but instead is updated numerous times per day.

```python
#Process Ratings Columns
dfnetflix.loc[dfnetflix["rating"].isnull()]
```

```python
# found out the rating from internet for this shows and fixing it.
rating_replacements = {
    67: 'TV-PG',
    2359: 'TV-14',
    3660: 'TV-MA',
    3736: 'TV-MA',
    3737: 'NR',
    3738: 'TV-MA',
    4323: 'TV-MA'
}

for id, rate in rating_replacements.items():
    dfnetflix.iloc[id, 8] = rate #8  here is the rating columns
    
dfnetflix['rating'].isnull().sum()
```

 It is difficult to understand the ratings associated with type of audiences as there are wide type so mapping rating in to 3 categories with help from Netflix site.  Not diving it in little kid and younger kid,as  I know both my kids in different age group watch same programmes together .And I assume same goes into many households, as we dont keep seperate TV for each kid.Teens contents are bit different so grouping as Kids,Teen,Adults.
```python
ratings_ages = {  'TV-Y' : 'Kids',
            'TV-Y7' : 'Kids' ,
            'TV-G'  : 'Kids',
            'PG'    : 'Kids',
            'TV-PG' : 'Kids',
            'PG-13' : 'Teens',
            'TV-14' : 'Teens',
            'R'     : 'Adults',
            'TV-MA' : 'Adults',
            'NC-17' : 'Adults',
            'NR': 'Adults',
            'UR': 'Adults',
            'TV-Y7-FV':'Kids',
            'G': 'Kids'} 
```

```python
dfnetflix['ratings_group'] = dfnetflix['rating'].replace(ratings_ages)
dfnetflix['ratings_group'].unique()
```

```python
#Process Duration Columns
# Result cols
df_duration = dfnetflix.copy()
dur_cols = ['title', 'duration', 'duration_num','type']

# Extracting numerical info for title duration
df_duration['duration_num'] = df_duration['duration'].apply(lambda x: int(x.split(' ')[0]))

# Results
df_duration.head().loc[:, dur_cols]
```

```python
# handling missing values in netflix data set
dfnetflix['director'] = dfnetflix['director'].fillna('NoDataAvailable')
dfnetflix['country'] = dfnetflix['country'].fillna(dfnetflix['country'].mode()[0])
dfnetflix['cast'] = dfnetflix['cast'].fillna('NoDataAvailable')
dfnetflix['date_added'] = dfnetflix['date_added'].fillna(dfnetflix['date_added'].mode()[0])
print(dfnetflix.isna().sum())
```

```python
dfnetflix.info()
```

```python
# adding few columns to dataset by extracting existing columns
dfnetflix['date_added'] = pd.to_datetime(dfnetflix.date_added)#convert to datetime
#adding month and year columns
dfnetflix['Month_Whenadded'] = dfnetflix['date_added'].dt.month
dfnetflix['Year_Whenadded'] = dfnetflix['date_added'].dt.year
dfnetflix['Month_name']= dfnetflix['date_added'].dt.month_name()

#how old is the content
dfnetflix['title_age_years'] = pd.datetime.now().year - dfnetflix['release_year']
# get num and season
dfnetflix['duration_num'] = dfnetflix['duration'].apply(lambda x: int(x.split(' ')[0]))
dfnetflix['duration_scale'] = dfnetflix['duration'].apply(lambda x: x.split(' ')[-1].strip())

# Fixing duration scale name season/Min
dfnetflix['duration_scale'] = dfnetflix['duration_scale'].apply(lambda x: 'Season' if x in ('Season', 'Seasons') else x)

# Sfinding day and month since added , rather than cmputing adding it has feature
dfnetflix['date_since_added'] = dfnetflix['date_added'].max()

# Computing deltas
dfnetflix['days_since_added'] = (dfnetflix['date_since_added'] - dfnetflix['date_added']).dt.days
dfnetflix['months_since_added'] = (dfnetflix['days_since_added'] / 30)
dfnetflix['years_since_added'] = (dfnetflix['days_since_added'] / 365)

dfnetflix=dfnetflix.loc[dfnetflix['date_added'] < '2021-06-27']  #creating subset of data only till 27 june 2021
dfnetflix.info()
```

<h2 style = "font-size : 25px ;font-weight:bold; color: #E50914">History and Evolution of Netflix</h2>
Netflix, Inc. is an American content platform and production company headquartered in Los Gatos, California.
Netflix was founded in 1997 by Reed Hastings and Marc Randolph in Scotts Valley, California.Netflix's initial business model included DVD sales and rental by mail, but Hastings abandoned the sales about a year after the company's founding to focus on the initial DVD rental business.Netflix expanded its business in 2007 with the introduction of streaming media while retaining the DVD and Blu-ray rental business.The company expanded internationally in 2010 with streaming available in Canada,[15] followed by Latin America and the Caribbean. This is an abridged history of the formation and growth of Netflix, which grown to become the largest entertainment company in the United States in terms of market capitalization as of 2021.Let look at Timeline of Netflix from DVD rentals to becoming most popular streaming services.

<br><h2 style="color:#E50914; font-weight:bold; font-family:TimesNewRoman">How does IMDB calculate the weighted average?</h2>

The following formula is used to calculate the Top Rated 250 titles. This formula provides a true 'Bayesian estimate', 
which takes into account the number of votes each title has received, minimum votes required to be on the list, 
and the mean vote for all titles:



$$weighted rating(wr)= \frac{v}{v+m}R+\frac{m}{v+m}C$$

Where:

R = average for the movie (mean) = (rating)

v = number of votes for the movie = (votes)

m = minimum votes required to be listed in the Top Rated list (currently 25,000)

C = the mean vote across the whole report

source :https://help.imdb.com/article/imdb/track-movies-tv/ratings-faq/G67Y87TFYYP6TWAV#

```python
def weighted_averages(number_votes, avg_vote, row):
    '''This function will calculate  the weighted avg'''
    if row[number_votes] != 0.0:
        weighted_rating = round(((row[number_votes]/(row[number_votes]+1000))*row[avg_vote]) + ((1000/(row[number_votes]+1000))*5.9),1)
    else:
        weighted_rating = 0
    return weighted_rating
```

<h2 style = "font-size : 25px ;font-weight:bold; color: #E50914">Top 10 Movies from year 2015 to 2021</h2>

```python
def find_plot_yearwise(fromyear,toyear, column, group):
    titles = []
    scores = []
    genre = []
    
    moviesfrom_to=df_netflix_imdb[(df_netflix_imdb['year']>fromyear) & (df_netflix_imdb['year']<=toyear)]
    top10=moviesfrom_to[column].nlargest(10)
    for i in range(len(top10)):
        index = top10.index[i]
        score = top10.iloc[i]
        scores.append(score)
        title = df_netflix_imdb['title'].iloc[index]
        titles.append(title)
    values=scores
    labels=titles
    
    fig, ax = plt.subplots(figsize=(13, 10), dpi=70)
    fig.patch.set_facecolor('black')
    ax.set_facecolor('black')
    
    y_pos = np.arange(len(labels))
    colors = ['#87050c','#b20710','#ba222a', '#ba222a','#c33e45','#c33e45','#cb595f','#d4757a','#dc9094','#e5acaf']
    ax.barh(y_pos, values, align='center',color=colors,alpha=0.9)
    ax.set_yticks(y_pos)
    ax.set_yticklabels(labels,{'fontsize': 14,'fontweight': 4,'color': 'white','verticalalignment': 'baseline'})
    ax.invert_yaxis()  # labels read top-to-bottom
    labels_score=values
    for loc in ['left','right','top','bottom']:
        ax.spines[loc].set_visible(False)
    # Text on the top of each bar
    i=0
    for p in ax.patches:
           x = p.get_x() + p.get_width()
           y = p.get_y() + p.get_height()
           ax.text(x,y, '{}'.format(labels_score[i]),{'font': 'serif', 'weight': 'normal', 'color': 'white', 'fontsize': 14}, alpha = 0.8)
           i=i+1 
    plt.grid(b=None,which='both', axis='both')
    ax.grid(b=None,axis='both',which='both')
    ax.axes.get_xaxis().set_visible(False)
    ax.set_title('Top 10  Rated  '+ 'from ' + str(fromyear) +' To '+  str(toyear) + " by " + group +"Rating" ,fontweight="bold",  fontsize=20, color='white')
    #ax.text(-58,10,'This visualization shows Top 10 countries .Vast content is released in US since Netflix was launched here',{'fontsize': 15,'fontweight': 9,'color': 'white','verticalalignment': 'baseline'})
```

```python
find_plot_yearwise(2015,2021,'weighted_average_vote','')
```
