# Base on our Historical Data we have Analyzed Let us Create a Model that will Predict the Educational Levels of the Students


```python
                               LOGISTIC REGRESSION
```

****

```python
dt=df[['gender','race/ethnicity','parental level of education','lunch','test preparation course']]
dt
```

```python
                           Data Preparation
```

```python
#                            Check Null VALUES
dt.isnull().count()
```

                    Lets Changes the Columns to 0 AND 1   

```python
dt['race/ethnicity']=dt['race/ethnicity'].apply(lambda x:'1' if x =='group A' else '2' if  x =='group B' else '3' if x =='group C' else '4' if x=='group D' else '5')
dt['gender']=dt['gender'].apply(lambda x: 0 if x =='male' else 1)
dt['lunch']=dt['lunch'].apply(lambda x: 1 if x=='standard' else 0)
dt['test preparation course']=dt['test preparation course'].apply(lambda x: 0 if x=='none' else 1)
dt['parental level of education']=dt['parental level of education'].apply(lambda x: 2 if x =="bachelor's degree" else 4 if x=="some college" else 1 if x == "master's degree" else 3)
dt
```

Our data is now prepared its ready for Prediiction. We will use Diffe****rents Models in order to check which model is best and scored highest Prediction accuracy

```python
x=dt.drop(columns=['parental level of education'])
x
```

```python
y=dt['parental level of education']
y
```

```python
L_model.fit(x,y)
```

```python
X_datatrain,X_datatest,y_datatrain,y_datatest = train_test_split(x,y, test_size=.85)
predic=L_model.predict(X_datatest)
acc = mean_squared_error(y_datatest,predic)
LOGISTIC_MODEL=acc*100
LOGISTIC_MODEL
```

```python
#                     RANDOM FOREST CLASSIFIER
```

```python
R_model.fit(x,y)
```

```python
X_datatrain,X_datatest,y_datatrain,y_datatest = train_test_split(x,y, test_size=.85)
predic=R_model.predict(X_datatest)
acc = mean_squared_error(y_datatest,predic)
R_FOREST_MODEL=acc*100
R_FOREST_MODEL
```

```python
#                     DECISION TREE
```

```python
D_model.fit(x,y)
```

```python
X_datatrain,X_datatest,y_datatrain,y_datatest = train_test_split(x,y, test_size=.85)
predic=D_model.predict(X_datatest)
acc = mean_squared_error(y_datatest,predic)
DEC_MODEL=acc*100
DEC_MODEL
```

```python
#                   KNEIGHBORS
```

```python
K_model.fit(x,y)
```

```python
X_datatrain,X_datatest,y_datatrain,y_datatest = train_test_split(x,y, test_size=.85)
predic=K_model.predict(X_datatest)
K_M= mean_squared_error(y_datatest,predic)
K_MODEL=K_M*100
K_MODEL
```

```python
from pandas import DataFrame 
from pandas import Series
```

```python
accuracy_scores=pd.DataFrame([LOGISTIC_MODEL,R_FOREST_MODEL,DEC_MODEL,K_MODEL],["Logist Model","Random_F Model","Decision Tree Model",
"KNeighbors_Model"],["ACCURACY SCORES OF THE MODELS"])
accuracy_scores
```
