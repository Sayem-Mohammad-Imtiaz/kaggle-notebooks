# Extracting the records where a team wo n batting first

```python
batting_first = ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head()
```

# Historgam about Win By Runs

```python
plt.figure(figsize=(9,9))
plt.hist(batting_first['win_by_runs'])
plt.title('Distribution of Runs')
plt.xlabel('Runs')
plt.show()
```

From above Histogram, we can say that most of the 'win_by_runs' matches won by  0-40 runs and less matches has been ended with margin of 100-140 runs (near 4-5 matches)

# Finding out the number of wins w.r.t. each team after batting first

```python
batting_first['winner'].value_counts()
```

# Top 3 teams with most wins after batting first (BarPlot)

```python
plt.figure(figsize=(9,9))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]), color=['blue', 'yellow', 'pink'])
plt.show()
```

# Making Pie Chart

```python
plt.figure(figsize=(10,10))
plt.pie(list(batting_first['winner'].value_counts()), labels=list(batting_first['winner'].value_counts().keys()), autopct ='%0.1f%%')
plt.show()
```
