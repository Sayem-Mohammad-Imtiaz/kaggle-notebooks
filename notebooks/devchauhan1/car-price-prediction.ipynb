```python
def plot_count(x,fig):
    plt.subplot(4,2,fig)
    plt.title(x+' Histogram')
    sns.countplot(cars[x],palette=("magma"))
    plt.subplot(4,2,(fig+1))
    plt.title(x+' vs Price')
    sns.boxplot(x=cars[x], y=cars.price, palette=("magma"))
    
plt.figure(figsize=(15,20))

plot_count('enginelocation', 1)
plot_count('cylindernumber', 3)
plot_count('fuelsystem', 5)
plot_count('drivewheel', 7)

plt.tight_layout()
```

Inference :

1.Very few datapoints for enginelocation categories to make an inference.

2.Most common number of cylinders are four, six and five. Though eight cylinders have the highest price range.

3.mpfi and 2bbl are most common type of fuel systems. mpfi and idi having the highest price range. But there are few data for 

4.other categories to derive any meaningful inference

5.A very significant difference in drivewheel category. Most high ranged cars seeme to prefer rwd drivewheel.

**Drop Unnecessary Columns**

Car ID, symboling, car name does not affect the price.

```python
cars=cars.drop(['car_ID','CompanyName','symboling'],axis=1)
```

```python
cars.head()
```

**APPLYING LABEL ENCODING**

```python
pd.set_option('display.max_columns', None)
cars.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df_col=("fueltype","aspiration","doornumber","carbody","drivewheel","enginelocation","enginetype","cylindernumber","fuelsystem")
```

```python
for i in df_col:
    cars[i]=le.fit_transform(cars[i])
```

```python
cars.head()
```

```python
cars.info()
```

```python
X=cars.iloc[:,:-1].values
y=cars.iloc[:,-1].values
```

```python
print(X)
```

**TRAIN TEST SPLIT OF DATA**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

**APPLYING RANDOM FOREST REGRESSION MODEL ON DATASET**

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X_train, y_train)
```

**PREDICTING RESULTS**

```python
y_pred=regressor.predict(X_test)
```

```python
regressor.score(X_test,y_test)
```

**ACCURACY SCORE**

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
