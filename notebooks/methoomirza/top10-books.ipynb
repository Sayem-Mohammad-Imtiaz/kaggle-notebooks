```python
data = pd.read_csv("/kaggle/input/gufhtugu-publications-dataset-challenge/GP Orders - 5.csv")
mydata = data
mydata.info()
mydata.head()
```

1. Replace name of headers for Easy to use

```python
mydata = mydata.rename({"Order Number": "OrdNumber", 
                        "Order Status": "OrdStatus", 
                        "Book Name" : "BookName", 
                        "Order Date & Time": "OrdDate", 
                        "City": "city", 
                        "Payment Method": "PayMethod", 
                        "Total items": "TotItems", 
                        "Total weight (grams)": "weight"}, 
                       axis = 1 )
mydata.head()
```

2. Null values

```python
mydata.isnull().sum()
```

3. Drop Null values and check 

```python
mydata = mydata.dropna()
mydata.isnull().sum()
```

4. Total Number of Books

```python
total_books = len(mydata['BookName'])
print("Total Number of Books are ", total_books)
```

5. in Single Order Row there are multiple Books Name Concatenate with ("/", "," and "aur" )
6. need to replace these string and characters with ("/")

```python
mydata.BookName = mydata.BookName.str.replace('aur','/')
mydata.BookName = mydata.BookName.str.replace(',','/')
```

```python
after_total_books = len(mydata['BookName'])
print("After replacing Total Number of Books are [", after_total_books,"] Same Numbers")
```

7. Now Apply the Split Function to Seprate each Book Name

```python
myBook_list = list(chain.from_iterable(mydata.BookName.str.split('/')))
after_split_tot_books = len(myBook_list)
print("After Split Function Total Number of Books are [", after_split_tot_books,"]")
```

8. Create DataFrame and Copy the List in
9. Confirm datatyoe as String (str) apply Function strip to remove spaces
10. value counts function for number if unique record count

```python
mybooks = pd.DataFrame(myBook_list, columns=['name'])
mybooks.name.astype('str')
mybooks.name.str.strip()
mybooks.name.value_counts()
```

11. List of Top Sold Books Name

```python
getData = mybooks["name"].value_counts()[:10].sort_values().to_frame()
print('list of Top 10 Sold Books Details\n----------------------------------\n',getData)
```

12. Representation of Top 10 Sold Books Name

```python
import plotly.express as px
px.bar(getData, y = getData.index, x = getData['name'], title = 'Top 10 Sold Books',
      labels={'index':'Books Name', 'name':'Number of Books'} )
```

```python
mybooks['name'].str.strip().value_counts().iloc[:20].plot.bar()
plt.title("Top 20 Sold Books", fontsize=18)
plt.xlabel("Name of Sold Books", fontsize=18)
plt.ylabel("Total Numbers of Books Sold", fontsize=12)
plt.tick_params(labelsize=15)
```

```python
get20Data = mybooks["name"].value_counts()[:20].sort_values().to_frame()
px.bar(get20Data, x = get20Data.index, y = 'name', title = 'Top 20 Sold Books',
      labels={'index':'Books Name', 'name':'Number of Books'} )
```
