```python
data = pd.read_csv('/kaggle/input/pakistanintellectualcapitalcs/Pakistan Intellectual Capital - Computer Science - Ver 1.csv', encoding="ISO-8859-1")
data.info()
```

```python
# REMOVING FIRST COLUMN (S#) AS NOT REQUIRED
mydata = data
mydata = mydata.drop('S#', axis=1)
mydata.head()
```

```python
mydata['Department'].value_counts()
```

```python
# REPLACE DEPARTMENT NAME IN DATASET FOR DATA UNIQUENESS
mydata['Department'] = mydata['Department'].replace('Computer Science', 'Computer Sciences')
mydata['Department'] = mydata['Department'].replace('Department of Computer Systems & Software Engineering', 'Computer Science and Software Engineering')
mydata['Department'] = mydata['Department'].replace('Computer Science & Engineering', 'Computer Science and Engineering')
mydata['Department'] = mydata['Department'].replace('CS/IT', 'Computer Science & Information Technology')
mydata['Department'] = mydata['Department'].replace('CS & IT', 'Computer Science & Information Technology')
mydata['Department'] = mydata['Department'].replace('Computer Science and IT', 'Computer Science & Information Technology')
mydata['Department'] = mydata['Department'].replace('Computer Science & IT', 'Computer Science & Information Technology')
mydata['Department'] = mydata['Department'].replace('Computer Science and Information Technology', 'Computer Science & Information Technology')
mydata['Department'] = mydata['Department'].replace('Computing & Information Sciences', 'Computer and Information Sciences')
```

```python
mydata['Department'].value_counts()
```

```python
mydata['Terminal Degree'].unique()
```

```python
mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("Ph.D","PhD")
mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("Phd","PhD")
mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("Ph.D (Scholar)","PhD")
mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("Ph.D(Scholar)","PhD")

mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("Mphil","M.Phil")
mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("MPhil","M.Phil")
mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("M.Phil leading to PhD","M.Phil")

mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("MSc","M.Sc")
mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("MSC","M.Sc")

mydata['Terminal Degree'] = mydata['Terminal Degree'].replace(" MCS","MCS")
mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("Mcs","MCS")

mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("MS (Scholar)","MS")
mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("M.S","MS")

mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("M.E","ME")

mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("BSC","B.Sc")
mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("BSc","B.Sc")

mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("B.E","BE")

mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("PostDoc","Post Doc")
mydata['Terminal Degree'] = mydata['Terminal Degree'].replace("Master","Masters")
```

```python
mydata['Terminal Degree'].unique()
```

```python
mydata["Country"].unique()
```

```python
mydata["Country"] = mydata["Country"].replace(" USA", "USA")
mydata["Country"] = mydata["Country"].replace(" New Zealand", "New Zealand")
mydata["Country"] = mydata["Country"].replace(" Germany", "Germany")
mydata["Country"] = mydata["Country"].replace(" Sweden", "Sweden")
```

```python
mydata["Country"].unique()
```

```python
mydata['Country'] = mydata['Country'].astype('str')
sorted(mydata['Country'].unique())
```

# Abundance Area of Interest / Expertise in Pakistan

```python
mydata['Area of Specialization/Research Interests'] = mydata['Area of Specialization/Research Interests'].str.strip()
mydata['Area of Specialization/Research Interests'] = mydata['Area of Specialization/Research Interests'].str.lower()
```

```python
print("Area of Interest \t\t\t\t\t\t Strength\n-------------------------------------------------------------------------\n",
      mydata['Area of Specialization/Research Interests'].str.strip().value_counts().iloc[:10] )
```

```python
plt.figure(figsize = (30, 8))
plt.style.use('seaborn-white')

plt.subplot(1,2,1)
ax=sns.countplot(x="Area of Specialization/Research Interests", hue="Province University Located", data=mydata,
              order=mydata['Area of Specialization/Research Interests'].value_counts().iloc[:10].index)

ax.set_xticklabels(ax.get_xticklabels(),fontsize=14,rotation=40, ha="right")
ax.set_title('Top 10 Interest/Expertise Area Provience Wise',fontsize= 22)
ax.set_xlabel('Area of Interest/Expertise Per Proviences',fontsize = 20) 
ax.set_ylabel('Number of Staff Members', fontsize = 20)
ax.legend(loc='upper right', frameon=False)
plt.tight_layout()
```

# Pakistan and Other Countries Professors Ratio

```python
mydata['Country'] = mydata['Country'].str.lower()
print ("Teachers/Faculty from Countries\n--------------------------------\n",
       mydata['Country'].value_counts().iloc[:10] )
```

```python
country_criteria = ["pakistan", "usa", "india", "china"]
dataCriteria = mydata[mydata["Country"].isin(country_criteria)]
print ("Faculty Members from Pakistan, USA and China not found from India\n-------------------------------------------------------------------\n",
       dataCriteria["Country"].value_counts() )
```

```python
country_criteria = ["pakistan", "usa", "india", "china"]
degree_criteria = ["PhD"]
dataCriteria2 = mydata[mydata["Country"].isin(country_criteria) & mydata["Terminal Degree"].isin(degree_criteria)]
print ("PhD Faculty Members from Pakistan, USA and China\n--------------------------------------------------\n",
       dataCriteria2["Country"].value_counts() )
```

```python
country_criteria = ["pakistan", "usa", "india", "china"]
degree_criteria = ["PhD"]
designation_criteria = ["Professor"]
dataCriteria2 = mydata[mydata["Country"].isin(country_criteria) & mydata["Terminal Degree"].isin(degree_criteria)
                       & mydata["Designation"].isin(designation_criteria)]
print ("PhD Faculty Members from Pakistan, USA and China With Designation [Professors]\n" +
       "-------------------------------------------------------------------------------\n",
       dataCriteria2["Country"].value_counts() )
```
