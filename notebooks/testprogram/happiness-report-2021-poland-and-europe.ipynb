# A function that reads data

```python
def Read():
    df = pd.read_csv('/kaggle/input/world-happiness-report-2021/world-happiness-report-2021.csv') 
    return df
```

# Column renaming function

```python
def RenamingColumns(df, columns: list):
    df.columns = col
    return df
```

```python
col = ["Kraj", "Wskaźnik regionalny", "Wynik drabinkowy1", "Wynik drabinkowy2",
       "Górny wąs", "dolny wąs", "PKB na mieszkańca", "Wspacie Społeczne", "Oczekiwana długość życia w zdrowiu",
       "Wolność dokonywania życiowych wyborów", 
       "Hojność",
       "Postrzeganie korupcji", "Drabina w dystopii",
       "Wyjaśnione przez: Logowanie PKB na mieszkańca",
       "Wyjaśnione przez: Wsparcie społeczne",
       "Wyjaśnione przez: Zdrowa długość życia",
       "Wyjaśnione przez: Wolność dokonywania życiowych wyborów",
       "Wyjaśnione przez: Szczodrość",
       "Wyjaśnione przez: Postrzeganie korupcji",
       "Dystopia + pozostałości"]
```

```python
df = Read()
df = RenamingColumns(df, col)
```

# # # Selecting interesting columns for analysis. 
I chose the following columns:
* Country, 
* Regional indicator, 
* GDP per capita, 
* Social Support, 
* Healthy Life Expectancy, 
* Ladder Outcome1

```python
def SelectingAColumn(df, columns: list):
    DF = df[columns]
    return DF
```

```python
columns = ['Kraj', "Wskaźnik regionalny", "PKB na mieszkańca", "Wspacie Społeczne",
           "Oczekiwana długość życia w zdrowiu",  "Wynik drabinkowy1"]
df = SelectingAColumn(df, columns)
```

**Selecting countries in Europe based on the region index. I was interested in Western Europe and Central and Eastern Europe**

```python
def daneforEurope(DF):
    DF = DF[(DF["Wskaźnik regionalny"] == "Western Europe") | (DF["Wskaźnik regionalny"] == "Central and Eastern Europe")]
    DF = DF.reset_index(drop=True)
    DF = DF.sort_values('Kraj')
    return DF
```

```python
df  = daneforEurope(df)
DF_Europe = df[['Kraj', "PKB na mieszkańca", "Wspacie Społeczne",
           "Oczekiwana długość życia w zdrowiu", "Wynik drabinkowy1"]]
```

## Selecting a few countries for further analysis

```python
def SelectingCountries(df, state):
    DF = df[df['Kraj'] == f"{state}"]
    return DF
```

```python
DF_Poland = SelectingCountries(DF_Europe,"Poland" ) 
DF_Finland = SelectingCountries(DF_Europe, "Finland")
DF_Denmark = SelectingCountries(DF_Europe, "Denmark")
DF_Estonia = SelectingCountries(DF_Europe,"Estonia")
DF_Hungary = SelectingCountries(DF_Europe, "Hungary")
Europe = pd.concat([DF_Poland,DF_Finland, DF_Denmark, DF_Estonia, DF_Hungary])
Europe = Europe.reset_index(drop=True)
Europe
```
