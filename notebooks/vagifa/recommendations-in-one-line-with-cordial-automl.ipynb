<h1> Recommendations in one line with Cordial AutoML </h1>

    Recently, I built an open-source AutoML library called Cordial. Cordial is
    an API that allows developers to easily run recommender systems.
    The library currently has 3 systems:
    1. BasicRecommender, this is the fastest but least accurate as it's quite
    primitive and uses linear kernels and matrix similarity
    2. LDARecommendeer, this uses linear discriminant analysis and euclidean 
    distance
    3. GraphRecommender, this uses complex graph networks to build a 
    recommender using networkx

    For this example, I have decided to use the GraphRecommender to demo the
    library

```python
!pip install cordial
```

```python
from cordial.content_recommenders import GraphRecommender
recommender = GraphRecommender('/kaggle/input/netflix-shows/netflix_titles.csv')
print(recommender.recommend('Zulu Man in Japan')['result'])
```

    As you can see, it takes the file path of the file as input on initialisation. Then,
    you call the recommend function, passing in the name of the movie you would like to
    get recommendations from ,and there you have it!
