# **Importing the relavent libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
data = pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
```

# **2 Cleaning the data**

```python
# check for the missing values 

data.isnull().sum()
```

```python
# find out the variables which are not important in deciding the output

student = data.drop(['race/ethnicity','parental level of education'],axis =1)
```

```python
student.head()
```

# **3 Analysis of relationship**

```python
#finding mutual relationship or connection between two or more variables

correlation = data.corr()
sns.heatmap(correlation,xticklabels=correlation.columns,yticklabels=correlation.columns,annot=True)
```

```python
sns.pairplot(student)
```

```python
student.head()
```

**using diffrent plotting methods understand the variable relationships** 

```python
sns.relplot(x='math score',y='reading score',hue='gender',data=student)
```

```python
sns.distplot(student['math score'])
```

```python
sns.distplot(student['reading score'])
```

```python
sns.distplot(student['writing score'])
```

```python
sns.catplot(x='math score',kind='box',data=student)
sns.catplot(x='reading score',kind='box',data=student)
sns.catplot(x='writing score',kind='box',data=student)
```
