Dapatkan dan `title`, `abstract`, serta `body text` paper tersebut

```python
# helper function untuk mendapatkan title berdasarkan document ID

id2title = []
for file in datafiles:
    with open(file,'r')as f:
        doc = json.load(f)
    id = doc['paper_id'] 
    title = doc['metadata']['title']
    id2title.append({id:title})
```

```python
# helper function untuk mendapatkan abstract berdasarkan document ID

id2abstract = []
for file in datafiles:
    with open(file,'r')as f:
        doc = json.load(f)
    id = doc['paper_id'] 
    abstract = ''
    if('abstract' in doc.keys()):
        for item in doc['abstract']:
            abstract = abstract + item['text']
        
    id2abstract.append({id:abstract})
```

```python
# helper function untuk mendapatkan body text berdasarkan document ID

id2bodytext = []
for file in datafiles:
    with open(file,'r')as f:
        doc = json.load(f)
    id = doc['paper_id'] 
    bodytext = ''
    for item in doc['body_text']:
        bodytext = bodytext + item['text']
        
    id2bodytext.append({id:bodytext})
```
