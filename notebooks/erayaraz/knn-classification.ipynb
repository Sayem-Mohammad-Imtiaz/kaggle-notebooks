# KNN Algorithm

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/kaggle/input/biomechanical-features-of-orthopedic-patients/column_2C_weka.csv')
df
```

```python
df.info()
```

```python
df.describe().T
```

```python
df["class"] = [ 1 if each == "Abnormal" else 0 for each in df["class"]]
df
```

```python
abnormal = df[df["class"] == 1]
normal = df[df["class"] == 0]

# scatter plot
plt.scatter(abnormal.sacral_slope,abnormal.pelvic_radius,color="red",label="Abnormal",alpha=0.5)
plt.scatter(normal.sacral_slope,normal.pelvic_radius,color="blue",label="Normal",alpha=0.5)
plt.xlabel("sacral_slope")
plt.ylabel("pelvic_radius")
plt.show()
```

1- Choose K value
2- Find the nearest data points in K
3- Calculate how many of the class nearest neighbors in K
4- Determine which class of point or data we tested belongs to

```python
y = df["class"].values
y
```

```python
x_data = df.drop(["class"],axis=1) # axis=1 for columns
x_data
```

# Normalization

Normalization: I represent the data to a value between 0 and 1 for more accurate processing. Normalization is very important process at KNN to  prevent deactivation due to the difference between large values and small values.
 

```python
x = (x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data))
x
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

# KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=13) # n_neighbors => key value
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
```

```python
prediction
```

I said an estimated 13. Here, a method can be used to find the n_neighbors value.



```python
print("{} Knn score: {} ".format(13,knn.score(x_test,y_test)*100))
```

```python
# find k value
score_list = []

for each in range(1,15):
    knn2 = KNeighborsClassifier(n_neighbors=each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))
    
plt.plot(range(1,15),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```
