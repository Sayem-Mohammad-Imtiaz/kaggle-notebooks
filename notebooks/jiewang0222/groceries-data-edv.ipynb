```python
grocery=pd.read_csv('/kaggle/input/groceries-dataset/Groceries_dataset.csv')
```

discover the dataset size and columns. There are three columns in this dataset:'Member_number'---member_number of customers, 'Date'--purchasing date of customer and 'itemDescription'--names of purchasing items. All three variables are without missing values. It is a relative clean dataset, but the type of column 'Date' should be redeemed.

```python
grocery.info()
```

```python
grocery['Date']=pd.to_datetime(grocery['Date'])
```

```python
date_diff=grocery['Date'].max()-grocery['Date'].min()
date_diff
```

According to the date difference of date, it is clear that this is a dateset for almost two years.

```python
print(grocery.head(20))
```

First I checked the acculated used amount of each member_number.

```python
gr=grocery.groupby('Member_number').count().sort_values(['Date','itemDescription'],ascending=False)

print(gr.head(5))
```

As shown above, the member_number'3188' is the customer, who purchased among the most frequency in last two years. Therefore, I add one bariable--'frequency',setting 'low' frequency for counts below 10,'middle' for below 20,'high' above 20.

```python
# create a list of our conditions
conditions = [
    (gr['Date']  > 0) & (gr['Date']  <=10),
    (gr['Date'] > 10) & (gr['Date'] <= 20),
    (gr['Date'] > 20)
    ]

# create a list of the values we want to assign for each condition
values = ['low','middle','high']

# create a new column and use np.select to assign values to it using our lists as arguments
gr['frequency']=np.select(conditions, values)
gr.head(5)
```

```python
import matplotlib.pyplot as plt
```

```python
gr1=gr.groupby('frequency').count()
gr1.sort_values('Date',ascending=False).plot.bar()
plt.title('counting for frequency')
plt.show()
```

```python
gr1.head(5)
```

It will be clearer by using pie chart to show the fraction of frequency.

```python
gr1['fraction']=gr1['Date']/gr1['Date'].sum()
# Creating plot 
fig = plt.figure(figsize =(10, 7)) 
plt.pie(gr1['fraction'], labels = gr1.index,autopct='%1.2f') 
plt.legend(title='Frequency') 
# show plot 
plt.show() 
```
