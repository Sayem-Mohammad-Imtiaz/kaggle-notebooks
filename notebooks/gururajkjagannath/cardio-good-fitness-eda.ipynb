### AGE, MILES

```python
# Customer Age
plt.figure(figsize = (15,5))
plt.subplot(1, 2, 1)                          #setting up subplots
sns.histplot(x=in_data.Age, kde=False, color='c')   #Plot histogram
plt.axvline(in_data.Age.mean(), color='g', linewidth=2, label='Mean')      #draw mean
plt.axvline(in_data.Age.median(), color='b', linestyle='dashed', linewidth=2, label='Median')  #Median
plt.axvline(in_data.Age.mode()[0], color='y', linestyle='--', linewidth=2, label='Mode')     #Mode
plt.title('Age Distribution', fontweight='bold')         #Chart title
plt.legend()    #Display the legend on Plots

# Miles 
plt.subplot(1, 2, 2)
sns.histplot(x=in_data.Miles, kde=False, color='c')
plt.axvline(in_data.Miles.mean(), color='g', linewidth=2,label='Mean')
plt.axvline(in_data.Miles.median(), color='b', linestyle='dashed', linewidth=2, label='Median')
plt.axvline(in_data.Miles.mode()[0], color='y', linestyle='--', linewidth=2, label='Mode')
plt.title('Miles Distribution', fontweight='bold')         #Chart title
plt.legend()

plt.rcParams["axes.labelsize"] = 15
plt.tight_layout()
plt.show()
```

# Bivariate Analysis


## PRODUCT
## VS
## AGE, EDUCATION, USAGE, FITNESS, INCOME, MILES

```python
#Min, Max, Mean(Average) of all the numerical variables by Product

kj=pd.pivot_table(in_data, 
               values=['Age','Education','Usage','Fitness','Income','Miles'], 
               index=['Product'],
               aggfunc={
                        'Age': [min, max, np.mean, np.median],
                        'Education': [min, max, np.mean, np.median],
                        'Usage': [min, max, np.mean, np.median],
                        'Fitness': [min, max, np.mean, np.median],
                        'Income': [min, max, np.mean, np.median],
                        'Miles': [min, max, np.mean, np.median]
                       }
               )
kj.T
```

```python
########################################################################
#PLOTTING [PRODUCT] V/S [AGE, EDUCATION, USAGE, FITNESS, INCOME, MILES]
########################################################################

figure, axes = plt.subplots(3, 2, figsize=(20,25))    #3 rows and 2 subplots in each
figure.suptitle('Product and relation with other numerical variables',fontsize=16,fontweight='bold') #title of the chart
k=[(0,0),(0,1),(1,0),(1,1),(2,0),(2,1)]  #Setting up indices to access axes
cols=['Age','Education','Usage','Fitness','Income','Miles']   #different fields to plot against Product
for i in (enumerate(cols)):        #assigns count to each of value in cols 
    sns.boxplot(
                x='Product', 
                y=i[1], 
                data=in_data, 
                ax=axes[k[i[0]][0]][k[i[0]][1]], 
                palette='Blues_r',
                showmeans=True,
                meanprops={"marker":"D",
                           "markerfacecolor":"blue", 
                           "markeredgecolor":"black",
                           "markersize":"7"
                          }
               ).set(title='Product by ' + i[1])  #Boxplot to see the distribution
    
plt.rcParams["axes.labelsize"] = 15
plt.show()
 
```

```python
#Product by fitness plot above shows that for Product TM195 and TM498, the 25, 50 & 75 quartiles are same
#lets check that out.

print('Product: TM195')
display(pd.crosstab(in_data['Product'],in_data['Fitness']).query("Product=='TM195'").quantile([0.25, 0.5, 0.75]))
print('\nProduct: TM498')
display(pd.crosstab(in_data['Product'],in_data['Fitness']).query("Product=='TM498'").quantile([0.25, 0.5, 0.75])) 
```
