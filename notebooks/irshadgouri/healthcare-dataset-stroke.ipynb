**CLEANING THE DATA**

```python
data.isnull().sum()
```

```python
data.bmi.fillna(data.bmi.mean(),inplace=True) 
```

```python
data.isnull().sum()
```

```python
print(data.work_type.value_counts())
print('-------------')
print(data.gender.value_counts())
print('-------------')
print(data.smoking_status.value_counts())
print('-------------')
print(data.Residence_type.value_counts())
```

```python
data.work_type.str.get_dummies('children')
```

```python
column=['gender', 'hypertension', 'heart_disease', 'ever_married',
       'work_type', 'Residence_type',
       'smoking_status']
column
```

```python
for i in column:
    data[i].value_counts().plot(kind='bar')
    plt.title(i,size=20)
    plt.grid()
    sns.set_theme(style='darkgrid')
    plt.show()
```

```python
sns.distplot(data['avg_glucose_level'])
```

```python
plt.figure(figsize=(10,5))
plt.hist(data.bmi, color='yellowgreen',edgecolor='k')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(data.avg_glucose_level, bins=30, color='yellowgreen' , alpha=0.6,edgecolor='k')
plt.hist(data.bmi, bins=30, color='#FF8F00' , alpha=0.8)
plt.show()
```
