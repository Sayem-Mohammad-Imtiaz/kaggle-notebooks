### 결측값을 보아하니 armed,age,gender,race,flee 열의 결측값이 있다는 것을 확인할 수 있습니다.

> 여기서 우리는 armed를 제외한 나머지 컬럼의 결측값을 채울 예정입니다.

```python
age = np.array(data['age'])
nan_age = np.where(np.isnan(data['age']))
nan_age = nan_age[0]
# 결측값 표시

mean_age = int(np.mean(data['age']))

for i in range(len(nan_age)):
    data['age'][nan_age[i]] = mean_age
```

나이의 결측값을 채우기 위해 일단 컬럼을 가져와 age라는 변수에 하나 저장합니다.

다음에는 넘파이 라이브러리 중 하나인 where함수를 통해서 결측값의 인덱스를 가져옵니다.

결측값을 채우는 방향은 모든 나이의 평균을 통해 집어 넣어줄 것이기 때문에 mean함수를 통해서 결측값 위치 인덱스의 나이의 평균을 넣어줍니다.

```python
Gender = np.array(data['gender'])
nan_gender = np.where(data['gender'].isnull())
nan_gender = nan_gender[0]
# 결측값 표시

for i in range(len(nan_gender)):
    data['gender'][nan_gender[i]] = 'M'
```

성별같은 경우에는 2개의 결측값만 존재하기 때문에 모두 남자로 채워주겠습니다. 

```python
data.isnull().sum()
```

위에 저 두 코드를 실행한 결과 결측값 처리가 되어있다는 것을 알 수 있습니다.

```python
race = set(data['race'])
race = list(race)
race
```

race의 결측값을 채우기위해서 race에는 어떤 값들이 있는지 판단하기 위해 set함수를 통해서 중복을 제거하여 알아냅니다.

```python
del race[0]
race
```

nan을 제거하기위해 del 이라는 명령어를 통해 제거해줍니다.

```python
Race = np.array(data['race'])
nan_race = np.where(data['race'].isnull())
nan_race = nan_race[0]
for i in range(len(nan_race)):
    data['race'][nan_race[i]] = race[i % 6]
```

Race에서도 똑같이 결측값을 채워주는데 여기서 결측값에 똑같은 인종이 들어가는 것은 아니라고 생각하여 결측값 순서대로 인종을 다양하게 넣어주었습니다.

```python
data.isnull().sum()
```

Race의 결측값이 없어진 형태

```python
flee = set(data['flee'])
flee = list(flee)
```

flee에서도 마찬가지로 결측값을 없애주기위해 flee에는 어떤 값들이 들어있는지 파악하기 위해 Race처럼 set함수를 통해서 원소값을 알아냅니다.

```python
flee
```

nan을 제외한 나머지가 4개있다는 것을 우리는 알게되었으니 nan을 제거해주도록 합니다.

```python
del flee[1]
flee
```

race와 마찬가지로 제거한 후에 모습을 나타냅니다.

```python
Flee = np.array(data['flee'])
nan_flee = np.where(data['flee'].isnull())
nan_flee = nan_flee[0]
for i in range(len(nan_flee)):
    data['flee'][nan_flee[i]] = [i % 4]
```

race의 결측값 채운 것처럼 flee도 똑같이 다양한 값을 넣어 편향이 없도록 하게 구현하였습니다.

armed를 제외한 나머지의 결측값이 없게 되었습니다.

```python
# 1. Shooting rate

shot_feature = set(data['manner_of_death'])
print(shot_feature)
shooting = np.where(data['manner_of_death'] == 'shot')
shot_and_Tasered = np.where(data['manner_of_death'] == 'shot and Tasered')

import seaborn as sns

sns.countplot(x="manner_of_death",data = data)
plt.show()
print("총을 쏜 비율 :" ,np.round(len(shooting[0]) / len(data['manner_of_death']),3))
print("총과 테이저를 쏜 비율 :" ,np.round(len(shot_and_Tasered[0]) / len(data['manner_of_death']),3))
```

우리는 첫번째 분석인 shooting rate를 분석하기 위해서 seaborn 라이브러리를 이용하여 countplot 함수를 가지고 데이터를 시각화합니다.

총을 쏜것도 총만 쐈는지 총과 테이저를 같이 쐈는지를 분석하기 위해서 데이터의 manner_of_death 컬럼을 가지고 와서 분석을 시작합니다.

위와 같은 코드를 작성하여 시각화한 결과입니다. 총만 쏜 비율은 95퍼 테이저건까지 같이한 경우는 5퍼에 해당됩니다.

```python
# What is the rate of killings relative to race and age

plt.figure(figsize=(10,5))
sns.countplot(x="race",data = data)
plt.show()

plt.figure(figsize=(15,5))
sns.distplot(data["age"])
plt.show()
```
