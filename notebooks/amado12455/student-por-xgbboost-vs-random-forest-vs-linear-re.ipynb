```python
data=df_not_out[["school","Fjob","studytime","internet","G1","G2","Medu","Fedu"]]
```

```python
data=pd.get_dummies(data)
```

```python
data
```

```python
medu=data.pop("Medu")

data["medu_0"] = (medu==0)*1

data["medu_1"] = (medu==1)*1

data["medu_2"] = (medu==2)*1

data["medu_3"] = (medu==3)*1

data["medu_4"] = (medu==4)*1
```

```python
fedu=data.pop("Fedu")

data["fedu_0"] = (fedu ==0)*1

data["fedu_1"] = (fedu ==1)*1

data["fedu_2"] = (fedu ==2)*1

data["fedu_3"] = (fedu ==3)*1

data["fedu_4"] = (fedu ==4)*1
```

```python
studytime=data.pop("studytime")

data["studytime_1"] = (studytime ==1)*1

data["studytime_2"] = (studytime ==2)*1

data["studytime_3"] = (studytime ==3)*1

data["studytime_4"] = (studytime ==4)*1
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scala=StandardScaler()
```

```python
data_scala=scala.fit_transform(data)
target=np.array(df_not_out.G3).reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(data_scala,
                                               target,
                                               test_size=0.2,
                                               random_state=0)
```

# **Linear Regression**

```python
from sklearn.linear_model import LinearRegression
```

```python
reg_lineal=LinearRegression()
```

```python
reg_lineal.fit(X_train,Y_train)
```

```python
reg_lineal.score(X_test,Y_test)
```

```python
reg_lineal.score(X_train,Y_train)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
y_pred=reg_lineal.predict(X_test)

mean_absolute_error(Y_test,y_pred)
```

# **RandomForest**

```python
from sklearn.ensemble import RandomForestRegressor

reg=RandomForestRegressor(max_depth=3)

reg.fit(X_train,Y_train)

reg.score(X_test,Y_test)

```

```python
mean_absolute_error(Y_test,reg.predict(X_test))
```

# **XGBOOST**

```python
from xgboost import XGBRegressor

xgb=XGBRegressor(max_depth=1)

xgb.fit(X_train,Y_train)

xgb.score(X_test,Y_test)
```

```python
xgb.score(X_train,Y_train)
```

```python
mean_absolute_error(Y_test,xgb.predict(X_test))
```

```python
xgb_pred=xgb.predict(X_test).flatten()
lineal_pred=reg_lineal.predict(X_test).flatten()
rm_pred=reg.predict(X_test).flatten()
```

```python
df_test=pd.DataFrame({"True values": Y_test.flatten() , "XGB predict": xgb_pred, "Random forest predict": rm_pred,
                      "Lineal Regression predict": lineal_pred })
                      
```
