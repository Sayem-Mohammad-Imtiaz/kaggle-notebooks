```python
%config Completer.use_jedi = False
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../input/unsupervised-learning-on-country-data/Country-data.csv')
df
```

```python
col_details=pd.read_csv('../input/unsupervised-learning-on-country-data/data-dictionary.csv')
pd.set_option('display.max_colwidth', None)
col_details
```

```python
country=df['country']
df.drop(['country'],inplace=True,axis=1)
```

```python
df_copy=df.copy()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df_copy)
```

```python
pd.DataFrame(scaled_data)
```

```python
from sklearn.cluster import KMeans
cost_values = []
for k in range(1, 15):
    model = KMeans(n_clusters=k)
    model.fit(scaled_data)
    cost_values.append(model.inertia_)

plt.plot(cost_values)
```

```python
k = 3
model = KMeans(n_clusters=k)
clusters = model.fit_predict(scaled_data)
```

```python
clusters
```

```python
df['country']=country
df['clusters']=clusters
df
```

```python
df.describe()
```

```python
plt.figure(figsize=(15,10))
plt.xticks([])
plt.scatter(df.country[df.clusters==0],df.child_mort[df.clusters==0],color='blue')
plt.scatter(df.country[df.clusters==1],df.child_mort[df.clusters==1],color='olive')
plt.scatter(df.country[df.clusters==2],df.child_mort[df.clusters==2],color='orchid')
plt.xlabel('Countries',size=16)
plt.ylabel('Child Mortality',size=16)
plt.legend(['Developing','Developed','Under-Developed'])
```
