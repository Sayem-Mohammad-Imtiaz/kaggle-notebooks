## 0. Data Preparation

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
books = pd.read_csv("../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv")

books.head()
```

## 1. Scatter Plot of User Rating and Price

```python
sns.set_style("whitegrid")
sns.set_context("talk")

g1 = sns.relplot(x = "User Rating",
            y = "Price",
            data = books,
            kind = "scatter",
            col = "Genre",
            col_order = ["Fiction", "Non Fiction"],
            hue = "Price")

g1.fig.suptitle("Relationship Between User Ratings and Prices", y = 1.07)

g1.set_titles("{col_name}")

plt.show()
```

## 2. Trends between the User Rating and Book Type in 10 Years

```python
sns.set_style("whitegrid")
sns.set_context("talk")

sns.relplot(x = "Year",
            y = "User Rating",
            data = books,
            kind = "line",
            hue = "Genre",
            ci = None,
            aspect = 1.3).fig.suptitle("Trends of User Rating (2010 - 2019)", y = 1.04)

plt.show()
```

## 3. Compare the User Ratings with Reviews in Different Book Type

```python
sns.set_style("whitegrid")
sns.set_context("talk")

g3 = sns.relplot(x = "User Rating",
            y = "Reviews",
           data = books,
           kind = "scatter",
           hue = "Price",
           size = "Price",
           col = "Genre",
           col_order = ["Fiction", "Non Fiction"],
           height = 6)


g3.fig.suptitle("Relationship Between Ratings and Reviews in Different Book Type", y = 1.05)

g3.set_titles("{col_name}")

g3.set(xlabel = "Ratings")

plt.show()
```

## 4. Compare User Rating Distribution between Different Book Type

```python
sns.set_style("whitegrid")
sns.set_context("paper")

sns.catplot(x = "Genre", 
            y = "User Rating",
            kind = "box",
            data = books,
            sym = "").fig.suptitle("Distribution of User Ratings in Different Book Type", x = 0.6, y = 1.03)  # Remove Outliers


plt.show()
```

## 5. Point Plots of User Rating and Reviews

```python
sns.set_style("whitegrid")
sns.set_context("talk")

sns.catplot(x = "User Rating", 
            y = "Reviews",
            kind = "point",
            data = books,
            hue = "Genre",
            aspect = 2.5,
            #join = False,
            ci = None,
            estimator = np.median,
            capsize = 0.1).fig.suptitle("Relationship between Book Ratings and Number of Reviews", y = 1.05)


plt.show()
```

## 6. Point Plots of User Rating and Prices

```python
sns.set_style("whitegrid")
sns.set_context("talk")

sns.catplot(x = "User Rating", 
            y = "Price",
            kind = "point",
            data = books,
            hue = "Genre",
            #join = False,
            ci = None,
            estimator = np.median,
            aspect = 2).fig.suptitle("Relationship between User Ratings and Book Prices", y = 1.05)


plt.show()
```
