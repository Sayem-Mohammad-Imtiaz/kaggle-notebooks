```python
df2=df.drop(["Row ID","Order ID","Customer ID","Product ID"],axis=1)
df2
```

```python
df2.isnull().sum()
```

```python
df2[df2["Postal Code"].isnull()]
```

```python
df2["Postal Code"]=df2["Postal Code"].fillna(5402)
```

```python
df2.isnull().sum()
```

```python
df2
```

```python
df2[df2["Postal Code"]==5402]
```

```python
df2["Mounth"]=df2["Order Date"].str[3:5]
df2["Mounth"]=df2["Mounth"].astype("int32")
df2
```

```python
d = {1:"January",2:"February",3:"March",4:"April",5:"May",6:"June",7:"July",
    8:"August",9:"September",10:"October",11:"November",12:"December"}
df2["Mounth"]=[d[i] for i in df2["Mounth"]]
df2
```

```python
df2["Order Date"] = pd.to_datetime(df2["Order Date"])
df2["Ship Date"] = pd.to_datetime(df2["Ship Date"])

df2
```

```python
df2["Year"]=df2["Order Date"].dt.year
df2
```

```python
df2["Year"].value_counts()
```

```python
mounths_df = df2.groupby("Mounth").sum()
mounths_df = mounths_df[["Sales"]]
mounths_df
```

```python
mounths_df.reset_index(inplace=True)
mounths_df
```

```python
plt.bar(mounths_df["Mounth"],mounths_df["Sales"])
plt.xlabel("Mounths")
plt.ylabel("Sales")
plt.xticks(fontsize=14,rotation=90)
plt.yticks(fontsize=14)
plt.show()
```

```python
year_df = df2.groupby("Year").sum()
year_df = year_df[["Sales"]]
year_df
```

```python
year_df.reset_index(inplace=True)
year_df
```

```python
plt.bar(year_df["Year"],year_df["Sales"])
plt.xlabel("Years")
plt.ylabel("Sales")
plt.show()
```

```python
customers = df2.groupby("Customer Name").sum().sort_values("Sales",ascending=False).head(20)
customers.drop(["Postal Code","Year"],axis=1,inplace=True)
customers
```

```python
customers.reset_index(inplace=True)
customers
```

```python
plt.bar(customers["Customer Name"],customers["Sales"])
plt.xlabel("Customer Name")
plt.ylabel("Sales")
plt.xticks(fontsize=12,rotation=90)
plt.yticks(fontsize=14)
plt.show()
```

```python
category = df2.groupby("Category").sum()
category.drop(["Postal Code","Year"],axis=1,inplace=True)
category
```

```python
category.reset_index(inplace=True)
category
```

```python
plt.bar(category["Category"],category["Sales"])
plt.xlabel("Category")
plt.ylabel("Sales")
plt.xticks(fontsize=12,rotation=90)
plt.yticks(fontsize=14)
plt.show()
```

```python
plt.pie(category["Sales"],labels=category["Category"],autopct="%0.2f%%",explode=[0,0,0.1])
plt.show()
```
