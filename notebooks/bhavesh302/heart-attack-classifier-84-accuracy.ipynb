```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
sns.set()
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
import matplotlib.pyplot as plt
```

```python
heart_df=pd.read_csv("/kaggle/input/heart-attack-analysis-prediction-dataset/heart.csv")
saturation_df=pd.read_csv("/kaggle/input/heart-attack-analysis-prediction-dataset/o2Saturation.csv")
```

```python
heart_df.head()
```

```python
vect = ['age','trtbps','chol','oldpeak','thalachh']
```

```python
for column_name in vect:
    scaler = StandardScaler()
    heart_df[column_name] = scaler.fit_transform(np.array(heart_df[column_name]).reshape(-1,1))
     
```

```python
target = heart_df['output']
heart_df = heart_df.drop('output',axis=1)
```

```python
from sklearn.model_selection  import train_test_split,cross_val_score
X_train,X_test,y_train,y_test = train_test_split(heart_df,target,test_size=0.20,random_state=42)
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lr_model  = LogisticRegression(n_jobs=-1,random_state=42)
lr_model.fit(X_train,y_train)
print("The cross validation score is: \n{}".format(np.std(cross_val_score(lr_model,X_train,y_train,cv=10))*100))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,lr_model.predict(X_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(n_estimators=150,n_jobs=-1,random_state = 42)
rf_model.fit(X_train,y_train)
print("The cross validation score is: \n{}".format(np.std(cross_val_score(rf_model,X_train,y_train,cv=10))*100))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,rf_model.predict(X_test)))
```

```python
#------------feature_importance_using_random_forest_model-----------
importances = rf_model.feature_importances_*100
```

```python
feature_list = X_train.columns
sample_df = pd.DataFrame([])
sample_df['features'] = feature_list
sample_df['importance_score'] = importances
```

```python
sample_df = sample_df.sort_values(by = ['importance_score'],ascending=False)
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='features',y='importance_score',data =sample_df)
```
