```python
data=pd.read_csv("/kaggle/input/league-of-legends-diamond-ranked-games-10-min/high_diamond_ranked_10min.csv")
```

```python
data.info()
```

Drop the gameId column.

```python
data_1 = data.drop(['gameId'],axis=1)
```

BlueKills and RedDeaths are same.
And drop other duplicated features.

```python
data_1 = data_1.drop(['blueGoldDiff','blueExperienceDiff','blueCSPerMin','blueGoldPerMin','redGoldDiff','redExperienceDiff','redCSPerMin','redGoldPerMin'],axis=1)
data_1 = data_1.drop(['redFirstBlood','redKills','redDeaths','blueAssists','redAssists'],axis = 1)
```

```python
data_1.info()
```

```python
data.head()
```

Normalize data by dividing 10000 ~ 10....

```python
data_1['blueTotalGold'] = data_1['blueTotalGold']/10000
data_1['blueTotalExperience'] = data_1['blueTotalExperience']/10000
data_1['redTotalGold'] = data_1['redTotalGold']/10000
data_1['redTotalExperience'] = data_1['redTotalExperience']/10000
data_1['blueTotalMinionsKilled'] = data_1['blueTotalMinionsKilled']/100
data_1['redTotalMinionsKilled'] = data_1['redTotalMinionsKilled']/100
data_1['blueTotalJungleMinionsKilled'] = data_1['blueTotalJungleMinionsKilled']/10
data_1['redTotalJungleMinionsKilled'] = data_1['redTotalJungleMinionsKilled']/10
data_1['blueWardsPlaced'] = data_1['blueWardsPlaced']/10
data_1['redWardsPlaced'] = data_1['redWardsPlaced']/10
```

split x,ytrain and x,ytest data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_1, data_1['blueWins'], test_size=0.30)
```

```python
X_train.head()
```

```python
Drop Bluewins(label) of X_train and X_test. 
```

```python
X_train = X_train.drop(['blueWins'],axis=1)
X_test = X_test.drop(['blueWins'],axis=1)
```

```python
X_train.head()
```

```python
X_test_m = X_test.to_numpy()
X_train_m = X_train.to_numpy()
y_test_m = y_test.to_numpy()
y_train_m = y_train.to_numpy()
```

```python
tf.model = tf.keras.Sequential()
tf.model.add(tf.keras.layers.Dense(input_dim = len(X_test_m[0]), units=1, activation = 'sigmoid'))
tf.model.compile(loss = 'binary_crossentropy',optimizer = tf.keras.optimizers.SGD(lr = 0.001),metrics = ['accuracy'])
tf.model.summary()

history = tf.model.fit(X_train_m, y_train_m, epochs=1000,batch_size=128)
```

Test the model

```python
print(tf.model.evaluate(X_test_m, y_test_m)[-1])
```
