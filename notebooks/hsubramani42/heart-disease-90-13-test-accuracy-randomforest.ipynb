# **Importing Data**

```python
import pandas as pd
data=pd.read_csv('../input/heart-disease-uci/heart.csv')
data.head()
```

### **Brief info about null and non-null for preprocessing**

```python
data.info()
```

## **Seperating feature columns from data**

```python
X=data[data.columns[data.columns!='target']]
X.head()
```

## **Seperating target column from data**

```python
Y=data['target']
Y.head()
```

## **Correlation b/w features and target**

```python
X.corrwith(Y)
```

## **Splitting data into train and split (ratio: 80:20)**

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y,test_y=train_test_split(X,Y,random_state=0,test_size=0.2)
print(train_y.shape, test_y.shape)
```

## **Using Random Forest with parameter tuning **(n_estimators,max_depth)****

```python
from sklearn.ensemble import RandomForestClassifier
best_estimator=0
best_depth=0
acc=0
for i in range(1,10):
    for j in range(1,10):
        classifier=RandomForestClassifier(n_estimators=i,max_depth=j,random_state=0,class_weight={0:sum(train_y==0),1:sum(train_y==1)})
        classifier.fit(train_x,train_y)
        if(classifier.score(test_x,test_y) > acc):
            best_estimator=i
            best_depth=j
            acc=classifier.score(test_x,test_y)
print('Best Estimator: ',best_estimator,'\nBest Estimator: ',best_depth,'\nAccuracy: ',acc)
```

## **Using best parameters**

```python
best_classifier=RandomForestClassifier(n_estimators=8,max_depth=5,random_state=0,class_weight={0:sum(train_y==0),1:sum(train_y==1)})
best_classifier.fit(train_x,train_y)
print('Train - Accuracy: ',round(best_classifier.score(train_x,train_y)*100,2),'%')
print('Test - Accuracy: ',round(best_classifier.score(test_x,test_y)*100,2),'%')
```

#### **Train and test acurracy are nearly same, so model is not overfitting or underfitting**

## **Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix
cfmt=confusion_matrix(test_y,best_classifier.predict(test_x))
print(cfmt)
```
