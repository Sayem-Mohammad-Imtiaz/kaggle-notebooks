````python
data=pd.read_csv("/kaggle/input/gufhtugu-publications-dataset-challenge/GP Orders - 4.csv")
````

````python
# remove null values 
data=data.dropna()
````

# **Spliting Data by month and Year**

````python
data['month_year'] = pd.to_datetime(data['Order Date']).dt.to_period('M')
data.head()
````

# **Selection Of Data**

````python
data=data[["Order Status","City (Billing)","month_year","Book Name"]]
````

````python
data = data.rename(columns={'Order Status': 'status',"City (Billing)":"city","Book Name":"name"})
````

````python
data.head()
````

# **change Status column to Numeric**

````python
status_class=[]
for i in data.status:
  if i =="Completed":
    status_class.append(1)
  elif i =="Returned":
    status_class.append(2)
  else:
    status_class.append(0)
````

````python
data["status_class"]=status_class
````

````python
data["status_class"].value_counts()
````

# **Data Visualization**

````python
unique_status=data.status.value_counts()
````

````python
unique_status
````

````python
unique_status.plot(kind="bar")
````

# **Impact of City On Orders**
