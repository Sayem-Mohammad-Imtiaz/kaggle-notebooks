# TRAIN TEST SPLIT

```python
x = df.drop(['output'],axis=1)
y = df['output']

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=0)

print('Train data records: %d \nTest data records: %d' % (x_train.shape[0], x_test.shape[0]))
```

# MODELLING 

```python
sc = StandardScaler()
sc.fit_transform(x_train)
sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state = 10)
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)

from sklearn.metrics import accuracy_score
a=accuracy_score(y_pred,y_test)
lrp=a*100
print("Percentage=",lrp)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier(random_state= 10)
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)

a=accuracy_score(y_pred,y_test)
dtp=a*100
print("Percentage=",dtp)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)

a=accuracy_score(y_pred,y_test)
knnp=a*100
print("Percentage=",knnp)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(random_state=1)
rf.fit(x_train,y_train)
```

```python
y_pred=rf.predict(x_test)

a=accuracy_score(y_pred,y_test)
rfp=a*100
print("Percentage=",rfp)
```

```python
from sklearn.svm import SVC
svc = SVC(probability=True)
svc.fit(x_train,y_train)
```

```python
y_pred=svc.predict(x_test)

a=accuracy_score(y_pred,y_test)
svcp=a*100
print("Percentage=",svcp)
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier(random_state=1)
xgb.fit(x_train,y_train)
```

```python
y_pred=xgb.predict(x_test)

a=accuracy_score(y_pred,y_test)
xgbp=a*100
print("Percentage=",xgbp)
```

```python
models=[lrp,dtp,knnp,rfp,svcp,xgbp]
plt.subplots(figsize=(15,5))
sns.countplot(models)
```

```python
print("Highest accuracy is:",max(models))
print("Model name - Logistic Regression Model")
```
