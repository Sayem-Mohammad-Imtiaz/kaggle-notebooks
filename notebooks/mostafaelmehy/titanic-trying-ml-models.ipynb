```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("../input/titanicdataset-traincsv/train.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sex_age_mean=df.groupby("Sex").Age.mean()
```

```python
df.loc[(df.Sex=="male")&(df["Age"].isnull()),"Age"]=sex_age_mean[1]
```

```python
df.loc[(df.Sex=="female")&(df["Age"].isnull()),"Age"]=sex_age_mean[0]
```

```python
df["Age"].isnull().sum()
```

```python
df["Embarked"].value_counts()
```

```python
df["Embarked"].fillna("S",inplace=True)
```

```python
df["Embarked"].isnull().sum()
```

```python
df.drop(["Cabin"],axis=1,inplace=True)
```

```python
df["Survived"].value_counts()
```

```python
df["Pclass"].value_counts()
```

```python
df["Sex"].value_counts()
```

```python
df["Sex"]=df["Sex"].map({"male":1,"female":0})
```

```python
df["Sex"].value_counts()
```

```python
df["Embarked"].value_counts()
```

```python
df=pd.get_dummies(df,columns=["Embarked"])
```

```python
df.head()
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
```

```python
df.columns
```

```python
X = df [['Pclass', 'Sex', 'Age', 'SibSp',
       'Parch', 'Fare', 'Embarked_C', 'Embarked_Q', 'Embarked_S']]
```

```python
y = df['Survived']
```

```python
clf = SVC(kernel='linear', C=1)
```

```python
scores = cross_val_score(clf, X, y, cv=5)
```

```python
scores
```

```python
scores.mean()
```

```python
scores.std()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
tuned_parameters = [{'kernel': ['rbf'], 'gamma': [1e-3, 1e-4],
                     'C': [1, 10, 100, 1000]},
                    {'kernel': ['linear'], 'C': [1, 10, 100, 1000]}]
```

```python
clf = GridSearchCV(
        SVC(), tuned_parameters,scoring='accuracy',
                     cv=5,
    )
```

```python
clf.fit(X,y)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
params = {'alpha': [0.01, 0.1, 0.5, 1.0, 10.0],}
```

```python
bernoulli_nb_grid = GridSearchCV(BernoulliNB(), param_grid=params, n_jobs=-1, cv=5, verbose=5,scoring='accuracy')
```

```python
bernoulli_nb_grid.fit(X,y)
```

```python
bernoulli_nb_grid.best_params_
```

```python
bernoulli_nb_grid.best_score_
```
