```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

%matplotlib inline
sns.set()
```

## Loading data

```python
data = pd.read_csv('../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
data.head()
```

## Reviewing the data

```python
data.info()
```

## Analysing the data

### Analysing data from Genre data

```python
data.Genre.value_counts()
```

```python
sns.countplot(x='Genre', data=data)
```

#### We can see that Non Fiction books have higher contribution in best seller

```python
plt.bar(x="Genre", height='Reviews', width=0.8, data=data)
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Reviews', fontsize=14)
plt.title('Number of reviews', fontsize=16)
plt.show()
data.groupby('Genre')['Reviews'].sum()
```

#### Even though Non fiction books are higher in number but Fiction books has higher consolidated reviews.

#### We can can say that Fiction books has demands and number of readers are higher

```python
Fiction = data[data['Genre']=='Fiction']
```

```python
Non_Fiction = data[data['Genre']=='Non Fiction']
```

```python
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.hist(Fiction['User Rating'], bins=10, color='black')
plt.xlabel('User Rating', fontsize=12)
plt.ylabel('Number of books', fontsize=12)
plt.title('Rating of Fiction books', fontsize=14)

plt.subplot(1,2,2)
plt.hist(Non_Fiction['User Rating'], bins=10)
plt.xlabel('User Rating', fontsize=12)
plt.ylabel('Number of books', fontsize=12)
plt.title('Rating of Non Fiction books', fontsize=14)

plt.show()
```

#### Rating of Fiction books are majorly consolidated from 4.6 and above rating and we see that there is a sudden spike of books having rating above 4.8.


#### In case of Non Fiction books we can see that all books are above Rating of 4.00 and majorly spike is there from rating 4.5 to 4.7

```python
data.groupby('Genre')['User Rating'].mean()
```

#### Fiction books has high average of overall User rating

```python
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.hist(data['Price'], bins=10, color='grey')
plt.xlabel('Price', fontsize=12)
plt.ylabel('Number of books', fontsize=12)
plt.title('Price range of all books', fontsize=14)
```

#### Majority of books have price 30 or below

```python
plt.figure(figsize=(14,7))
plt.subplot(1,2,1)
plt.hist(Fiction['Price'], bins=10, color='black')
plt.xlabel('Price', fontsize=12)
plt.ylabel('Number of books', fontsize=12)
plt.title('Price range of Fiction books', fontsize=14)

plt.subplot(1,2,2)
plt.hist(Non_Fiction['Price'], bins=10)
plt.xlabel('Price', fontsize=12)
plt.ylabel('Number of books', fontsize=12)
plt.title('Price range of Non Fiction books', fontsize=14)

plt.show()
```

```python
table = go.Figure(data=[go.Table(header=dict(values=['Books title', 'Author', 'Genre']),
         cells=dict(values=[data[data['User Rating']>=4.9]["Name"], data[data['User Rating']>=4.9]["Author"], data[data['User Rating']>=4.9]["Genre"]]))])
table.update_layout(title = 'Book which are rated 4.9',)
table.show()
```

```python
sns.regplot(x='Reviews', y='User Rating', data=data, ci=None)
plt.show()
```

#### There is no causuality effect of review on User Rating.
