```python
def fillage(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    return Age
```

```python
df['Age']=df[['Age','Pclass']].apply(fillage,axis=1)
```

**It is used to fill the null values of age based on the mean value we got from the box plot
   1 Pclass =37
   2 Pclass=29
   3 Pclass=24
**

```python
sns.heatmap(df.isnull(),yticklabels=False,cbar=False)
```

**Now we can see that there are no null values in the Age column**

**The name,sex ,ticket and embark are of object types.We can remove nameand ticket column and convert the sex and embark values into numerical format**

```python
sex=pd.get_dummies(df['Sex'],drop_first=True)
embark=pd.get_dummies(df['Embarked'],drop_first=True)
```

```python
df.drop(['Sex','Embarked','Ticket','Name'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df=pd.concat([df,sex,embark],axis=1)
```

# Logistic Regression****

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('Survived',axis=1)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,df['Survived'],test_size=0.3,random_state=101)
```

```python
xtrain.shape
```

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(xtrain,ytrain)
```

```python
ypred=reg.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
con=confusion_matrix(ytest,ypred)
print(con)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(ytest,ypred)
print(acc)
```
