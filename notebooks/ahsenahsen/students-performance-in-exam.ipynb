```python
def get_grade_for_exam(marks):
    if marks >= 85:
        return 'A'
    elif marks >= 70 and marks < 85:
        return 'B'
    elif marks >= 55 and marks < 70:
        return 'C'
    elif marks >= 40 and marks < 55:
        return 'D'
    elif marks < 40:
        return 'E'
    
df['reading_grade'] = df['reading'].apply(get_grade_for_exam)
df['writing_grade'] = df['writing'].apply(get_grade_for_exam)
df['math_grade'] = df['math'].apply(get_grade_for_exam)
df["allexam_mean_grade"] = df[["math","reading","writing"]].mean(axis=1).apply(get_grade_for_exam)
```

```python
df['allexam_pass'] = df.apply(lambda x : False if x['math_grade'] == False or 
                                    x['reading_pass'] == False or x['writing_pass'] == False else True, axis =1)
```

## MATH Score Prediction

```python
from sklearn.model_selection import RepeatedKFold,cross_validate, train_test_split
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, mean_squared_log_error
import re
```

```python


def evaluateRegressor(y, y_pred):
    
    MSE = mean_squared_error(y, y_pred)
    MAE = mean_absolute_error(y, y_pred )
    RMSE = mean_squared_error(y, y_pred, squared = False)
    MSLE = mean_squared_log_error(y, y_pred)
    R_squared = r2_score(y, y_pred )
    
    print("MSE:", MSE)
    print("MAE:", MAE)
    print("RMSE:", RMSE)
    print("MSLE:", MSLE)
    print("R-squared:", R_squared)
```

```python


df_train = df[["gender","race","PLE","lunch","TPC","math"]].copy()
df_train = pd.get_dummies(df_train)

# Bazı modellerde kolon isimleri ile alakalı hatalar ortaya çıkabiliyor. Bunu engelemek için kolon isimlerine regex uygulayarak uygun hale getiriyoruz.
df_train = df_train.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))

X = df_train.drop(["math"], axis=1)

Y = df_train["math"]

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = .2)
```

```python
from xgboost import XGBRegressor

regressor = XGBRegressor()

regressor.fit(x_train,y_train)

evaluateRegressor(y_test, regressor.predict(x_test))
```

```python
scoring=['r2','neg_root_mean_squared_error',"neg_mean_squared_error",'neg_mean_absolute_error']
cv = RepeatedKFold(n_splits=5, n_repeats=5)
scores = cross_validate(regressor,X,Y,cv=cv, scoring=scoring)
```

```python
print("R2: ",scores["test_r2"].mean())
print("MSE: ",(scores["test_neg_mean_squared_error"]* -1).mean())
print("RMSE: ",(scores["test_neg_root_mean_squared_error"]* -1).mean() )
print("MAE: ",(scores["test_neg_mean_absolute_error"] * -1).mean() )
```

```python
from sklearn.model_selection import RandomizedSearchCV
params = { 'max_depth': [3, 5, 6, 10, 15, 20],
           'learning_rate': [00.1,0.01, 0.1, 0.2, 0.3],
           'subsample': np.arange(0.5, 1.0, 0.1),
           'colsample_bytree': np.arange(0.4, 1.0, 0.1),
           'colsample_bylevel': np.arange(0.4, 1.0, 0.1),
           'n_estimators': [100, 500, 1000]}
xgbr = XGBRegressor(seed = 20, objective="reg:squarederror")
clf = RandomizedSearchCV(estimator=xgbr,
                         param_distributions=params,
                         scoring='neg_mean_squared_error',
                         n_iter=25,
                         verbose=1)
clf.fit(X, Y)
print("Best parameters:", clf.best_params_)
print("Lowest MSE: ", (-clf.best_score_))
```
