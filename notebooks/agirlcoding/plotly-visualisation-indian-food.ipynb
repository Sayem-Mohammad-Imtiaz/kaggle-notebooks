# Distribution of Dishes per Region

```python
fig=px.bar(indianFood, x='region'     
      )
#change color of the bins
fig.update_traces(marker_color='green',width=0.6)
fig.update_layout(title_text='<i><b>Number of Dishes per Region</b></i>', title_x=0.5, title_font_color="#043927",title_font_size=36, font_family="Courier New",
    font_color="#0B6623", xaxis_title="Indian Region",
    yaxis_title="Number of different Dishes", font_size=15)

fig.add_annotation( # add a text callout with arrow
    text="<b>New Region<br> in India? :) </b>", x="3", y=20, arrowhead=1, showarrow=True
)
```

# Which Region allowes for more Vegetarian Food?

```python
fig2=px.bar(
    indianFood, x='region', color= 'diet',
       title= 'Different Diet per Region', 
       labels= {'region': 'Region', 'diet': 'Diet'},
       color_discrete_sequence=['#3CB043','#32612D']
      )
fig2.update_layout(title_text='<i><b>Average Cooking Time per Dish per Region and Diet</b></i>', title_x=0.5, title_font_color="#03C04A",title_font_size=24, font_family="Courier New",
    font_color="#0B6623", xaxis_title="Indian Region",
    yaxis_title="Number of different Dishes", font_size=15)
```

```python
fig3=px.pie(indianFood, names='diet', color_discrete_sequence=['#3CB043','#32612D'])
fig3.update_traces(textfont_size=20, pull=[0.2, 0],
                  marker=dict(line=dict(color='#000000', width=2)))
fig3.update_layout(title_text='<i><b>Vegetarian or not Vegetarian?</b></i>', title_x=0.5, title_font_color="#03C04A",title_font_size=36, font_family="Courier New", font_size=15)
```

```python
col_one_list = indianFood['ingredients'][0].split(",")

ingredientsList=[]
for i in range(0,255):
    ingredientsList.append(indianFood['ingredients'][i].split(","))
    
flat_list = []
for sublist in ingredientsList:
    for item in sublist:
        flat_list.append(item)
        
countIngredients=[]
for i in flat_list:
   countIngredients.append(flat_list.count(i))

flat_small=map(lambda x:x.lower(), flat_list)
Ingredients_df = pd.DataFrame(list(zip(flat_small, countIngredients)), 
               columns =['Ingredient', 'Occurrence'])
sorted=Ingredients_df.sort_values(by=['Occurrence'], ascending=False)

unique_ingredients=sorted.drop_duplicates()
```
