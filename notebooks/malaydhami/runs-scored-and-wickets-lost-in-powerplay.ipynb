```python
import plotly.express as px

df1 = pd.read_csv("/kaggle/input/ipl-complete-dataset-20082020/IPL Ball-by-Ball 2008-2020.csv")
df2 = pd.read_csv("/kaggle/input/ipl-complete-dataset-20082020/IPL Matches 2008-2020.csv",parse_dates=['date'])
```

```python
# Merge both dataframes to get data by year
df2['year']=df2['date'].dt.year
df = pd.merge(df1,df2[["id","year"]],on="id")
df.head()
```

```python
# Dataframe contains only powerplay data
power_play = df.query("over < 6")
power_play = power_play.replace('Rising Pune Supergiant','Rising Pune Supergiants')
```

# **Total Powerplay runs,wickets**

```python
x = pd.pivot_table(df2,index="year",values="id",aggfunc="count").rename(columns={"id":"Total_matches"})
y = pd.pivot_table(power_play,index="year",values="batsman_runs",aggfunc="sum").rename(columns={"batsman_runs":"Total_runs"})
z = pd.pivot_table(power_play,index="year",values="is_wicket",aggfunc="sum").rename(columns={"is_wicket":"Total_wickets"})
data = pd.concat([x,y,z],axis=1)
data
```

# **Powerplay runs by each team (2008-2020)**

```python
fig = px.histogram(power_play,x="year",y="total_runs",histfunc="sum",hover_data={'batting_team':False},facet_col="batting_team",facet_col_wrap=3,height=1500,template='ggplot2')
fig.update_layout(bargap=0.2)
fig.update_xaxes(visible=False)
fig.update_yaxes(visible=False)
fig
```
