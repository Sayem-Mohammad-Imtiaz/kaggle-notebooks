<b> Null Values <b>

```python
wur.isnull().sum()
```

From the results above, we can see that there are 200 missing vlaues in the broad_impact column. Thus, we will fill the null values with 0 and check again for null values.

```python
wur['broad_impact'] = wur['broad_impact'].fillna(0)
```

```python
wur.isnull().sum()
```

<font size="3"> <b> Correlation and Heatmap </font> <b>

```python
wur.corr()
```

```python
corr = wur.corr(method = 'spearman')
plt.figure(figsize=(14,7))
plt.title("Correlation Graph")

sns.heatmap(corr, annot=True)
```

From the heatmap above, we can see that world ranking has a strong positive correlation with publications,broad impact, citations and influence.
