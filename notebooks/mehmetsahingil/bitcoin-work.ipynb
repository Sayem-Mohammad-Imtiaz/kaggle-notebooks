```python
df = pd.read_csv("../input/bitcoin-historical-data/bitstampUSD_1-min_data_2012-01-01_to_2021-03-31.csv");
df.info();
print(df.head());
```

```python
series = df["Weighted_Price"]; # Tek boyutlu olduğu için seridir.
print(type(series));
df_WP = df[["Weighted_Price"]]; # Bu durumda series olmaz, data frame olur
print(type(df_WP));
```

```python
timeStamp = df["Timestamp"];
timeStamp.plot(kind = "line");
print(timeStamp[0:20]);

# Drop timestamp
df = df.drop("Timestamp", axis = 1);
df.info();
```

```python
# Line plot of Open

#df.plot(kind = "line");
open = df["Open"];
open.plot(kind = "line", grid = "True", color = "red", alpha = 0.5, linestyle = ":");
plt.title("Open");
plt.xlabel("Indices");

plt.ylabel("Values");
```

```python
# Line plot of Volume_(BTC)
Volume_BTC = df["Volume_(BTC)"];
Volume_BTC.plot(kind = "line", color = "blue", alpha = 0.5, linestyle = ":");
plt.title("Volume_(BTC)");
plt.xlabel("Indices");
plt.ylabel("Values");
```

```python
# Line plot of Volume_(Currency)
Volume_Currency = df["Volume_(Currency)"];
Volume_Currency.plot(kind = "line", color = "red", alpha = 0.5, linestyle = ":");
plt.title("Volume_(Currency)");
plt.xlabel("Indices");
plt.ylabel("Values");
```

```python
# Correlation plot

corr = df.corr();
#print(corr);
f, ax = plt.subplots(figsize = (18, 18));
sns.heatmap(corr, annot = True, linewidth = 0.5, fmt = ".1f", ax = ax);
```

```python
# Scatter plot of Weghted Price vs Volume(BTC)
df.plot(kind = "scatter", x = "Weighted_Price", y = "Volume_(BTC)", alpha = .5, color = "purple");
plt.title("Weghted Price vs Volume(BTC)");
```

```python
# Scatter plot of Weghted Price vs Volume(Currancy)
df.plot(kind = "scatter", x = "Weighted_Price", y = "Volume_(Currency)", alpha = .5, color = "purple");
plt.title("Weghted Price vs Volume(Currancy)");
```

```python
# Change the name of Volume_(Currency) to VolumeCurrency
df["VolumeCurrency"] = df["Volume_(Currency)"];
df.drop("Volume_(Currency)", axis = 1);
print(df.columns);

# Plot histogram of VolumeCurrency
df.VolumeCurrency.plot(kind = "hist", bins = 30, figsize = (5, 5));
plt.title("Volume Currency Histogram");
```
