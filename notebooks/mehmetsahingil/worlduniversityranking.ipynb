```python
timesData = pd.read_csv("../input/world-university-rankings/timesData.csv");
timesData.info()
```

```python
timesData.head(10)
```

```python
df = timesData.iloc[:100, :];

import plotly.graph_objs as go
trace1 = go.Scatter(
                    x = df.world_rank,
                    y = df.citations,
                    mode = "lines",
                    name = "citations",
                    marker = dict(color = 'rgb(16, 112, 2, 0.8)'),
                    text = df.university_name);

trace2 = go.Scatter(
                    x = df.world_rank,
                    y = df.teaching,
                    mode = "lines+markers",
                    name = "teaching",
                    marker = dict(color = 'rgb(80, 26, 80, 0.8)'),
                    text = df.university_name);

data = [trace1, trace2];
layout = dict(title = "Citation and Teaching vs World Rank of Top 100 Universities",
             xaxis = dict(title = "World Rank", ticklen = 5, zeroline = False));

fig = dict(data = data, layout = layout);
iplot(fig)
```

```python
cond1 = timesData.year == 2014;
cond2 = timesData.year == 2015;
cond3 = timesData.year == 2016;
df2014 = timesData[cond1].iloc[:100, :];
df2015 = timesData[cond2].iloc[:100, :];
df2016 = timesData[cond3].iloc[:100, :];

trace1 = go.Scatter(
                    x = df2014.world_rank,
                    y = df2014.citations,
                    mode = "markers",
                    name = "2014",
                    marker = dict(color = 'rgb(16, 52, 2, 0.8)'),
                    text = df2014.university_name);

trace2 = go.Scatter(
                    x = df2015.world_rank,
                    y = df2015.citations,
                    mode = "markers",
                    name = "2015",
                    marker = dict(color = 'rgb(116, 112, 20, 0.8)'),
                    text = df2015.university_name);

trace3 = go.Scatter(
                    x = df2016.world_rank,
                    y = df2016.citations,
                    mode = "markers",
                    name = "2016",
                    marker = dict(color = 'rgb(156, 72, 90, 0.8)'),
                    text = df2016.university_name);

data = [trace1, trace2, trace3];
layout = dict(title = "Citation vs World Rank of Top 100 Universities in 2014, 2015 and 2016",
             xaxis = dict(title = "World Rank", ticklen = 5, zeroline = False),
             yaxis = dict(title = "Citation", ticklen = 5, zeroline = False));

fig = dict(data = data, layout = layout);
iplot(fig)
```

```python
cond = timesData.year == 2014;
df2014 = timesData[cond].iloc[:3, :];

trace1 = go.Bar(
                x = df2014.university_name,
                y = df2014.citations,
                name = "citations",
                marker = dict(color = 'rgba(16, 52, 2, 0.8)', 
                              line = dict(color = 'rgb(0,0,0)', width = 1.5)),
                text = df2014.country);

trace2 = go.Bar(
                    x = df2014.university_name,
                    y = df2014.teaching,
                    name = "teaching",
                    marker = dict(color = 'rgba(116, 152, 20, 0.8)', 
                                 line = dict(color = 'rgb(0,0,0)', width = 1.5)),
                    text = df2014.country);

data = [trace1, trace2];

#Group type barplot
layout = dict(title = "Citation & Teaching for Top 3 Universities in 2014",
             xaxis = dict(title = "Universities", ticklen = 5, zeroline = False),
             yaxis = dict(title = "Citation", ticklen = 5, zeroline = False),
             barmode = "group");

fig = dict(data = data, layout = layout);
iplot(fig)

#Relative type barplot
layout = dict(title = "Citation & Teaching for Top 3 Universities in 2014",
             xaxis = dict(title = "Universities", ticklen = 5, zeroline = False),
             yaxis = dict(title = "Citation", ticklen = 5, zeroline = False),
             barmode = "relative");

fig = dict(data = data, layout = layout);
iplot(fig)
```

```python
cond = timesData.year == 2016;
stRate = timesData[cond].iloc[:7, :];
pie1 = stRate.num_students;
pie1_list = [float(each.replace(',','.')) for each in pie1];
labels = stRate.university_name;

fig = {
    "data":[
        {
            "values": pie1_list,
            "labels": labels,
            "domain": {'x': [0, .5]},
            "name": "Number of students rates",
            "hoverinfo": "label+percent+name",
            "hole": .3,
            "type": "pie"
        },
    ],
    "layout": {
        "title": "Number of students for Universisities"        
    }
}

iplot(fig);
```

```python
df2011 = timesData.student_staff_ratio[timesData.year == 2011];
df2012 = timesData.student_staff_ratio[timesData.year == 2012];

trace1 = go.Histogram(x = df2011,
                     name = "2011",
                     opacity = .75);

trace2 = go.Histogram(x = df2012,
                     name = "2012",
                     opacity = .75);

data = [trace1, trace2];
layout = go.Layout(barmode = "overlay",
         title = "2011 vs 2012",
         xaxis = dict(title = 'Student staff ratio'),
         yaxis = dict(title = 'Count'));
fig = go.Figure(data = data, layout = layout);
iplot(fig);
```

```python
x2015 = timesData[timesData.year == 2015];

trace0 = go.Box(y = x2015.total_score,
                name = "Total score of universities in 2015",
                marker = dict(color = "rgb(12, 21, 41)"));

trace1 = go.Box(y = x2015.research,
                name = "Research of universities in 2015",
                marker = dict(color = "rgb(12, 128, 141)"));

data = [trace0, trace1];
iplot(data);
```

```python
x2015 = timesData[timesData.year == 2015];
data2015 = x2015.loc[:, ["research", "international", "total_score"]];
data2015["index"] = np.arange(1, len(data2015)+1);
fig = ff.create_scatterplotmatrix(data2015, diag = "box", index = "index", colormap = "Portland",
                                 colormap_type = "cat", height = 700, width = 700);
iplot(fig);
```

```python
trace1 = go.Scatter(
x = x2015.world_rank,
y = x2015.teaching,
name = "teaching",
marker = dict(color = "rgba(12,221,33,0.8)"));

trace2 = go.Scatter(
x = x2015.world_rank,
y = x2015.income,
xaxis = "x2",
yaxis = "y2",
name = "income",
marker = dict(color = "rgba(122,21,123,0.8)"));

layout = go.Layout(
xaxis2 = dict(domain = [0.6, 0.95], anchor = "y2"),
yaxis2 = dict(domain = [0.6, 0.95], anchor = "x2"));   

data = [trace1, trace2];
fig = go.Figure(data = data, layout = layout);
iplot(fig);
```

```python
# create trace 1 that is 3d scatter
trace1 = go.Scatter3d(
    x=x2015.world_rank,
    y=x2015.research,
    z=x2015.citations,
    mode='markers',
    marker=dict(
        size=x2015.teaching,
        color='rgb(255,0,0)',                # set color to an array/list of desired values      
    )
)

data = [trace1]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0  
    )
    
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
trace1 = go.Scatter(
    x=x2015.world_rank,
    y=x2015.research,
    name = "research"
)
trace2 = go.Scatter(
    x=x2015.world_rank,
    y=x2015.citations,
    xaxis='x2',
    yaxis='y2',
    name = "citations"
)
trace3 = go.Scatter(
    x=x2015.world_rank,
    y=x2015.income,
    xaxis='x3',
    yaxis='y3',
    name = "income"
)
trace4 = go.Scatter(
    x=x2015.world_rank,
    y=x2015.total_score,
    xaxis='x4',
    yaxis='y4',
    name = "total_score"
)
data = [trace1, trace2, trace3, trace4]
layout = go.Layout(
    xaxis=dict(
        domain=[0, 0.45]
    ),
    yaxis=dict(
        domain=[0, 0.45]
    ),
    xaxis2=dict(
        domain=[0.55, 1]
    ),
    xaxis3=dict(
        domain=[0, 0.45],
        anchor='y3'
    ),
    xaxis4=dict(
        domain=[0.55, 1],
        anchor='y4'
    ),
    yaxis2=dict(
        domain=[0, 0.45],
        anchor='x2'
    ),
    yaxis3=dict(
        domain=[0.55, 1]
    ),
    yaxis4=dict(
        domain=[0.55, 1],
        anchor='x4'
    ),
    title = 'Research, citation, income and total score VS World Rank of Universities'
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
