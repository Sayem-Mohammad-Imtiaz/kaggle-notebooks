```python
df1 = pd.read_csv('../input/hackathon/task_2-owid_covid_data-21_June_2020.csv')
df1.head()
```

```python
yemen1= df1[(df1['location']=='Yemen')].reset_index(drop=True)
yemen1.head()
```

```python
plt.figure(figsize=(10, 5))
ax=sns.countplot(x="date", data=yemen1,palette="Greens",edgecolor="black", order = yemen1['date'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Yemen Covid-19 by date')
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x()+0.25, p.get_height()+1), va='bottom',
                    color= 'black')
plt.show()
```

```python
fig = px.bar(yemen1, 
             x='aged_65_older', y='total_cases', color_discrete_sequence=['crimson'],
             title='Yemen Covid19 cases by 65y older', text='cvd_death_rate')
fig.show()
```

```python
fig = px.bar(yemen1, 
             x='gdp_per_capita', y='life_expectancy', color_discrete_sequence=['#27F1E7'],
             title='Yemen GDP per capita vs. Life expectancy', text='total_cases')
fig.show()
```

```python
fig, ax = plt.subplots(1,3, figsize = (20,6), sharex=True)
sns.countplot(x='total_cases',data=yemen1, palette="copper", ax=ax[0])
plt.xticks(rotation=45)
sns.countplot(x='total_cases',hue='aged_65_older', palette="ocean", data=yemen1,ax=ax[1])
plt.xticks(rotation=45)
sns.countplot(x='total_cases',hue='diabetes_prevalence', palette="cubehelix", data=yemen1,ax=ax[2])
ax[0].title.set_text('Yemen Covid19 totalCases')
ax[1].title.set_text('Yemen Covid19 Cases Vs 65y Older')
ax[2].title.set_text('Yemen Covid19 Cases Vs Diabetes Prevalence')
plt.xticks(rotation=45)
plt.show()
```

```python
#word cloud
from wordcloud import WordCloud, ImageColorGenerator
text = " ".join(str(each) for each in yemen1.location)
# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=200,colormap='Set2', background_color="black").generate(text)
plt.figure(figsize=(10,6))
plt.figure(figsize=(15,10))
# Display the generated image:
plt.imshow(wordcloud, interpolation='Bilinear')
plt.axis("off")
plt.figure(1,figsize=(12, 12))
plt.show()
```
