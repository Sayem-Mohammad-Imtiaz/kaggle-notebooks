#Codes from Paul Mooney (visualization of most common words)

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
def plotWordFrequency(input):
    f = open(Bulgaria,'r')
    words = [x for y in [l.split() for l in f.readlines()] for x in y]
    data = sorted([(w, words.count(w)) for w in set(words)], key = lambda x:x[1], reverse=True)[:40] 
    most_words = [x[0] for x in data]
    times_used = [int(x[1]) for x in data]
    plt.figure(figsize=(20,10))
    plt.bar(x=sorted(most_words), height=times_used, color = 'red', edgecolor = 'black',  width=.5)
    plt.xticks(rotation=45, fontsize=18)
    plt.yticks(rotation=0, fontsize=18)
    plt.xlabel('Most Common Words:', fontsize=18)
    plt.ylabel('Number of Occurences:', fontsize=18)
    plt.title('Most Commonly Used Words: %s' % (Bulgaria), fontsize=24)
    plt.show()
```

```python
Bulgaria = '../input/hackathon/task_1-google_search_txt_files_v2/BG/Bulgaria-en-result-113-original.txt'
plotWordFrequency(Bulgaria)
```

```python
df2 = pd.read_csv('../input/hackathon/task_2-owid_covid_data-21_June_2020.csv')
df2.head()
```

```python
bulgaria3 = df2[(df2['location']=='Bulgaria')].reset_index(drop=True)
bulgaria3.head()
```

```python
plt.figure(figsize=(10, 5))
ax=sns.countplot(x="tests_units", data=bulgaria3,palette="Greens",edgecolor="black", order = bulgaria3['tests_units'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Covid19 Tests Units')
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x()+0.25, p.get_height()+1), va='bottom',
                    color= 'black')
plt.show()
```

```python
fig = px.bar(bulgaria3, 
             x='aged_65_older', y='total_cases', color_discrete_sequence=['crimson'],
             title='Bulgarian Covid19 cases by 65y older', text='cvd_death_rate')
fig.show()
```

```python
fig = px.bar(bulgaria3, 
             x='gdp_per_capita', y='life_expectancy', color_discrete_sequence=['#27F1E7'],
             title='Bulgaria GDP per capita vs. Life expectancy', text='total_cases')
fig.show()
```

```python
fig, ax = plt.subplots(1,3, figsize = (20,6), sharex=True)
sns.countplot(x='total_cases',data=bulgaria3, palette="copper", ax=ax[0])
sns.countplot(x='total_cases',hue='aged_65_older', palette="ocean", data=bulgaria3,ax=ax[1])
sns.countplot(x='total_cases',hue='diabetes_prevalence', palette="cubehelix", data=bulgaria3,ax=ax[2])
ax[0].title.set_text('Covid19 totalCases')
ax[1].title.set_text('Covid19 Cases Vs 65y Older')
ax[2].title.set_text('Covid19 Cases Vs Diabetes Prevalence')
plt.show()
```
