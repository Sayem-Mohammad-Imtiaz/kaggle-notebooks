Venezuela is no longer a country with standards of living standards in South America. The level of poverty it has reached in recent years, and which has continued to grow between 2019 and 2020, puts the country, which has the largest reserves of world oil , on a different scale. These are the conclusions of the most recent Venezuelan Living Conditions Survey (Encovi, in the Spanish acronym), carried out by a group of universities annually since 2014 and considered the most accurate x-ray of the country's social situation since the government stopped to publish statistics.


After a decade of oil boom, ended in 2014, today the country is the poorest in the region and the second most unequal - behind only Brazil - according to calculations made from interviews in almost 10,000 households. The data reflects the economic collapse, in addition to the demographic and social adjustment caused by the forced emigration of five million Venezuelans.


Analyzing other aspects, such as employment, education, housing conditions and public services, poverty reaches 65 per cent of Venezuelan households - having increased by 13.8 per cent between 2018 and 2019.

The research included a new indicator called consumption poverty, which points out that 68 per cent of the population consumes less than the amount considered necessary for food per day, 2,000 calories. Food insecurity and precarious services have become an equalizing factor. Among the poorest and richest sectors, there are no notable differences in these aspects.

Although higher-income citizens consume five times more meat than the poorest, the majority of the population eats below the recommended protein per day: 51 grams. Only 32 per cent of the population manages to ingest the necessary. To this factor is added the anthropometric data on child malnutrition: 30 per cent of children under the age of five - about 639 thousand people - have chronic malnutrition or short stature. The percentage is much higher than that of neighboring countries, being only lower than that of Guatemala.

In terms of poverty and malnutrition, the country has acquired conditions that are similar to those of others in Central America, the Caribbean and Africa. https://translate.google.com.br/translate?hl=en&sl=pt&u=https://oglobo.globo.com/mundo/venezuela-79-da-populacao-esta-em-extrema-pobreza-mostra-estudo-24521940&prev=search&pto=aue

The composition of the Venezuelan population has also changed in recent years. There are less than four million inhabitants in the country, and in 19 per cent  of households someone emigrated between 2014 and 2019. Another of the study's conclusions confirms that the so-called demographic bonus has been lost. This factor is the result of a reduction in the fertility rate - families having fewer children - and a decrease in the mortality of a population - that is, people living longer. As a result, there is a greater proportion of citizens of working age - between 15 and 64 years old - in relation to the dependent population, children and the elderly.


All efforts to expand mass access to university education have lost their meaning. These benefits do not reach the poorest sectors.

#Venezuelan refugees receive humanitarian assistance in Brazil

In Brazil, Operation Welcome welcomes Venezuelan refugees and migrants who cross the border through the State of Roraima. This is the Humanitarian Logistics Task Force, carried out and coordinated by the Federal Government, with support from the Armed Forces, United Nations agencies and more than 100 civil society entities.

And, in a time of pandemic, solidarity makes the difference. Refugees came together to give back work. Beneficiaries of Operation Welcomed joined Brazilian seamstresses and tailors to produce Personal Protective Equipment (PPE), essential items to face the virus.
