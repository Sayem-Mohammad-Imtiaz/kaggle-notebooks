## Train-Val-Test

```python
x=x.values
y=y.values
```

```python
type(x)
```

```python
type(y)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.20)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,random_state=0,test_size=0.10)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_val.shape
```

```python
y_val.shape
```

## Standart Scaler

```python
from sklearn.preprocessing import StandardScaler

obje_ss=StandardScaler()

x_train_ss=obje_ss.fit_transform(x_train) 
x_val_ss=obje_ss.transform(x_val)
x_test_ss=obje_ss.transform(x_test)
```

```python
x_train_ss.shape
```

```python
x_train_ss
```

## Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,classification_report

model_logistic=LogisticRegression().fit(x_train_ss,y_train)
y_pred_logistic_val=model_logistic.predict(x_val_ss)

print('Logisticregression classifcation train başarı oranı %',model_logistic.score(x_train_ss,y_train)*100)
print(confusion_matrix(y_val,y_pred_logistic_val))
print('Logisticregression classifcation validation başarı oranı %',accuracy_score(y_val,y_pred_logistic_val)*100)
print('Logisticregression classifcation validation precision_score %',precision_score(y_val,y_pred_logistic_val)*100)
print('Logisticregression classifcation validation recall_score %',recall_score(y_val,y_pred_logistic_val)*100)
print('Logisticregression classifcation validation f1_score %',f1_score(y_val,y_pred_logistic_val)*100)
print(classification_report(y_val,y_pred_logistic_val))
```

```python
from sklearn.svm import SVC

model_svc=SVC(kernel='poly',degree=4).fit(x_train_ss,y_train)
y_pred_val_svc=model_svc.predict(x_val_ss)
print('svc classifcation train başarı oranı %',model_svc.score(x_train_ss,y_train)*100)
print(confusion_matrix(y_val,y_pred_val_svc)) 
print('svc classifcation validation başarı oranı %',accuracy_score(y_val,y_pred_val_svc)*100)
print('SVC classifcation validation precision_score %',precision_score(y_val,y_pred_val_svc)*100)
print('SVC classifcation validation recall_score %',recall_score(y_val,y_pred_val_svc)*100)
print('SVC classifcation validation f1_score %',f1_score(y_val,y_pred_val_svc)*100)
print(classification_report(y_val,y_pred_val_svc))
```

```python
from sklearn.ensemble import RandomForestClassifier

model_tree=RandomForestClassifier(n_estimators=50,max_depth=60,random_state=0).fit(x_train_ss,y_train)
y_pred_val_tree=model_tree.predict(x_val_ss)
print('tree classifcation train başarı oranı %',model_tree.score(x_train_ss,y_train)*100)
print(confusion_matrix(y_val,y_pred_val_tree))
print('tree classifcation validation başarı oranı %',accuracy_score(y_val,y_pred_val_tree)*100)
print('tree classifcation validation precision_score %',precision_score(y_val,y_pred_val_tree)*100)
print('tree classifcation validation recall_score %',recall_score(y_val,y_pred_val_tree)*100)
print('tree classifcation validation f1_score %',f1_score(y_val,y_pred_val_tree)*100)
print(classification_report(y_val,y_pred_val_tree))
```

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski').fit(x_train_ss,y_train)
y_pred_val_knn=model_knn.predict(x_val_ss)
print('KNN classifcation train başarı oranı %',model_knn.score(x_train_ss,y_train)*100)
print(confusion_matrix(y_val,y_pred_val_knn))
print('KNN classifcation validation başarı oranı %',accuracy_score(y_val,y_pred_val_knn)*100)
print('KNN classifcation validation precision_score %',precision_score(y_val,y_pred_val_knn)*100)
print('KNN classifcation validation recall_score %',recall_score(y_val,y_pred_val_knn)*100)
print('KNN classifcation validation f1_score %',f1_score(y_val,y_pred_val_knn)*100)
print(classification_report(y_val,y_pred_val_knn))
```

```python
from sklearn.naive_bayes import BernoulliNB,MultinomialNB

model_Bnb=BernoulliNB().fit(x_train_ss,y_train)
y_pred_val_Bnb=model_Bnb.predict(x_val_ss)
print('BernoulliNB classifcation train başarı oranı %',model_Bnb.score(x_train_ss,y_train)*100)
print(confusion_matrix(y_val,y_pred_val_Bnb))
print('BernoulliNB classifcation validation başarı oranı %',accuracy_score(y_val,y_pred_val_Bnb)*100)
print('BernoulliNB classifcation validation precision_score %',precision_score(y_val,y_pred_val_Bnb)*100)
print('BernoulliNB classifcation validation recall_score %',recall_score(y_val,y_pred_val_Bnb)*100)
print('BernoulliNB classifcation validation f1_score %',f1_score(y_val,y_pred_val_Bnb)*100)
print(classification_report(y_val,y_pred_val_Bnb))
```

```python
model_Mnb=MultinomialNB().fit(x_train,y_train)
y_pred_val_Mnb=model_Mnb.predict(x_val)
print('MultinomialNB classifcation train başarı oranı %',model_Mnb.score(x_train,y_train)*100)
print(confusion_matrix(y_val,y_pred_val_Mnb))
print('MultinomialNB classifcation validation başarı oranı %',accuracy_score(y_val,y_pred_val_Mnb)*100)
print('MultinomialNB classifcation validation precision_score %',precision_score(y_val,y_pred_val_Mnb)*100)
print('MultinomialNB classifcation validation recall_score %',recall_score(y_val,y_pred_val_Mnb)*100)
print('MultinomialNB classifcation validation f1_score %',f1_score(y_val,y_pred_val_Mnb)*100)
print(classification_report(y_val,y_pred_val_Mnb))
```

```python
import xgboost as xgb

model_xgb=xgb.XGBClassifier(objective='binary:hinge',max_depth=5,learning_rate=1).fit(x_train_ss,y_train)
y_pred_val_xgb=model_xgb.predict(x_val_ss)
print('XGB classifcation train başarı oranı %',model_xgb.score(x_train_ss,y_train)*100)
print(confusion_matrix(y_val,y_pred_val_xgb))
print(classification_report(y_val,y_pred_val_xgb))
```

```python
from sklearn.ensemble import AdaBoostClassifier

model_adaboost=AdaBoostClassifier(n_estimators=100,learning_rate=1.8).fit(x_train_ss,y_train)
y_pred_val_adaboost=model_adaboost.predict(x_val_ss)
print('Adaboost classifcation train başarı oranı %',model_adaboost.score(x_train_ss,y_train)*100)
print(confusion_matrix(y_val,y_pred_val_adaboost))
print(classification_report(y_val,y_pred_val_adaboost))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model_gradientboost=GradientBoostingClassifier(learning_rate=1.5,n_estimators=8).fit(x_train_ss,y_train)
y_pred_val_gradientboost=model_gradientboost.predict(x_val_ss)
print('Gradientboost classifcation train başarı oranı %',model_gradientboost.score(x_train_ss,y_train)*100)
print(confusion_matrix(y_val,y_pred_val_gradientboost))
print(classification_report(y_val,y_pred_val_gradientboost))
```

```python
import tensorflow as tf

model_c=tf.keras.Sequential()
model_c.add(tf.keras.layers.Dense(12, input_dim=12, activation='relu'))
model_c.add(tf.keras.layers.Dense(6, activation='relu'))
model_c.add(tf.keras.layers.Dense(1, activation='sigmoid'))
```

```python
model_c.compile(optimizer='Adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model_c.fit(x_train_ss,y_train,epochs=200,verbose=0)
```

```python
model_c.summary()
```

```python
y_pred_nn=model_c.predict(x_val_ss)
```

```python
y_pred_nn[y_pred_nn<0.5]=0
```

```python
y_pred_nn[y_pred_nn>0.5]=1
```
