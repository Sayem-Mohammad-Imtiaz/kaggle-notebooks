```python
# 11. Covert a column to row name/index in Pandas
housing = pd.read_csv ('../input/california-housing-prices/housing.csv')
df = pd.DataFrame(housing) 
print (df)
df2 = df.sample(10)
print (df2)
df3  = df2 [['total_rooms','median_income','population' ]]
print (df3)
df3.pivot(index='median_income', columns='total_rooms',values = 'population')
df3.pivot(index='population', columns='median_income')
```

```python
# 12. randomly select rows in a pandas dataframe
# selecting 10 rows from the dataframe
housing.sample (10)
```

```python
# Ranking Rows of Pandas DataFrame
housing['Ranking'] = housing['median_house_value'].rank(ascending = 1)
housing = housing.set_index('Ranking')
housing
```

```python
# Sorting rows in pandas DataFrame
# Sort the rows of dataframe by single column 'households'
housing = housing.set_index('households')
sortResult = housing.sort_values(by ='households', ascending = False)
sortResult
```

```python
# Sorting rows in pandas DataFrame
# Sort the rows of dataframe by multiple columns 'households' and 'population'
housing = housing.sort_values(by =['households', 'population'])
housing
```

```python
# Select row with maximum and minimum value in Pandas data frame
# Shows maximum on all attributes
housing.max()
```

```python
# Which row has maximum population in the dataframe
housing[housing.population == housing.population.max()]
```

```python
# Shows the Maximum population from the whole dataset
housing.population.max()
```

```python
# Shows minimum on all attributes
housing.min()
```

```python
# Which row has minimum population in the dataframe
housing[housing.population == housing.population.min()]
```

```python
# Shows the Minimum population from the whole dataset
housing.population.min()
```
