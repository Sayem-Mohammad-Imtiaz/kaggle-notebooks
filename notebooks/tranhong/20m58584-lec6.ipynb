# The country will be chosen to analyze is France. 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

selected_country_1 = 'France'
selected_country_2 = 'France'

df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1 = df1[df1['country']==selected_country_1]
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2 = df2[df2['country']==selected_country_1]
df1.index = pd.to_datetime(df1['date'])
df2.index = pd.to_datetime(df2['date'])
```

```python
print(df1.columns)
print(df2.columns)
```

```python
print(np.unique(df1['country']))
print(np.unique(df2['country']))
```

# Analyses the daily new cases and vaccinations
From here, I will analyze the datathat was loaded as an dataframe object

```python
fix,ax = plt.subplots(figsize=(10,5))
plt.plot(df1.index,df1['daily_new_cases'],label='Cases')
ax1 = ax.twinx()
plt.plot(df2.index,df2['daily_vaccinations'],'r',label='Vaccinations')
#plt.legend()
plt.show()
```

# Improved vissualization
From here, we will use the modules iplot and plotly to make visuallizations

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_new_cases = go.Scatter(x=df1.index,y=df1['daily_new_cases'],yaxis='y1',name='Daily confirmed cases')
daily_vaccinations = go.Scatter(x=df2.index,y=df2['daily_vaccinations'],yaxis='y2',name='Daily vaccinations')

layout_obj = go.Layout(title = 'COVID vs Vaccines',xaxis=dict(title='Date'),yaxis=dict(title='Daily Cases'),yaxis2=dict(title='Vaccinations',side='right',overlaying ='y'))
fig = go.Figure(data=[daily_new_cases,daily_vaccinations],layout = layout_obj)
iplot(fig)
fig.write_html('outcome.html')
```

# Global ranking of France
The index is added and start from 1 to determine the ranking of France in global scale.

```python
country = 'France'
df3 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df3 = df3.groupby('country').sum()['daily_new_cases'].sort_values(ascending=False)[0:10]
df3 = df3.reset_index()
df3.index = df3.index + 1
index1 = df3.loc[df3['country']==country]
print('The global ranking of',country, 'of the total cases is:')
print(index1)

df4 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df4 = df4.groupby('country').max()['people_fully_vaccinated_per_hundred'].sort_values(ascending=False)
df4 = df4.reset_index()
df4.index = df4.index + 1
index2 = df4.loc[df4['country']==country]
print('The global ranking of',country, 'of the total vaccinations per number of hundred people is:')
print(index2)

print(df3)
print(df4)
```
