## Little Description About our Happiness Data-Set.
<b>Life Ladder</b>: Imagine a ladder, with steps numbered from 0 at the bottom to 10 at the top. The top of the ladder represents the best possible life for you and the bottom of the ladder represents the worst possible life for you.

<b>Log GDP per capita</b>: At its most basic interpretation, per capita GDP shows how much economic production value can be attributed to each individual citizen. Alternatively, this translates to a measure of national wealth since GDP market value per person also readily serves as a prosperity measure.

<b>Social support</b>: Social support is defined in terms of social network characteristics such as assistance from family, friends, neighbours and other community members.

<b>Healthy life expectancy at birth</b>: Healthy life expectancy is the average life in good health - that is to say without irreversible limitation of activity in daily life or incapacities - of a fictitious generation subject to the conditions of mortality and morbidity prevailing that year

<b>Freedom to make life choices</b>: Freedom of choice describes an individual's opportunity and autonomy to perform an action selected from at least two available options, unconstrained by external parties   

<b>Generosity</b>: Generosity is the virtue of being liberal in giving, often as gifts.

<b>Perceptions of corruption</b>: Corruption is a form of dishonesty or criminal offense undertaken by a person or organization entrusted with a position of authority, to acquire illicit benefit

### Fun Part is Just going to Start, So get Ready.
First we will find Country Code with `country_name_to_country_alpha2`  by passing one arguemnt that is Country name than we will use `country_alpha2_to_continent_code` to find respective continent code with Country Code we find previous like AS for Asia, EU for Europe and so on.

```python
def get_continent(col):
    try:
        cn_a2_code = country_name_to_country_alpha2(col)
    except:
        cn_a2_code = 'Unknown'
    
    try:
        cn_continent = country_alpha2_to_continent_code(cn_a2_code)
    except:
        cn_continent = 'Unknown'
    
    return (cn_a2_code, cn_continent)

happiness_report ['Codes'] = happiness_report['Country name'].apply(get_continent)
happiness_report['Country Code'] = happiness_report ['Codes'].apply(lambda x: x[0])
happiness_report['Continent Code'] = happiness_report['Codes'].apply(lambda x : x[1])
happiness_report.drop('Codes',axis = 1, inplace = True)

## For Mapping Continent with their Full Name
continent_name = dict(AF = 'Africa', AS = 'Asia', EU = 'Europe',
                      NA = 'North America', OC = 'Oceania',
                      SA = 'South America', Unknown = 'Unknown')
## You can User map passing Dict which contain all Value you want to and it will to that for you
happiness_report['Continent Name'] = happiness_report['Continent Code'].map(continent_name)
happiness_report.head(5)
```

#### Now we Have Country Name, Country Code, Continent Name, Continent Code, So we Can Easily Find Corrdinate with Geopy which is python module getting address and Latitude, longitude

```python
geolocator = Nominatim(user_agent="myapplication")
def geolocate(country):
    try:     # You can use geolocator.reverse() which is opposit of this method you pass Latitude, longitude
        location = geolocator.geocode(country)
        return location
    except:
        return 'Unknown'
```

```python
## We Have 166 unique Country so we will run loop 166 time and find all Location.
country = happiness_report['Country name'].unique()
# In geopy there is no country as Taiwan Province of China so we have to remove it and run it
# and then add it longitudal and latitude mannuly (which you can find in google).
country = np.delete(country, 145)

lat = dict()
long = dict()

#This Loop will take time so we use tqdm which show progress bar and we can see it to know the progress.
for i in tqdm(range(len(country))):
    # Getting Lattiude and Longitude
    location = geolocate(country[i])
    lat[country[i]] = location.latitude
    long[country[i]] = location.longitude

lat['Taiwan Province of China'] = 23.6978
long['Taiwan Province of China'] =  120.9605

## Mapping Country Name with lattidue and Longitude
happiness_report['Latitude'] = happiness_report['Country name'].map(lat)
happiness_report['Longitude'] = happiness_report['Country name'].map(long)
```

```python
# Now Grouping Our dataset on basis of Country Name
happiness_report_mean = happiness_report.groupby('Country name').mean().reset_index()
happiness_report_mean.head(5)
```

```python
## initializing Our Map with Some Argumenet.
m = folium.Map(location = [25.90499783974841, 74.65802895310858],tiles='Stamen Terrain', zoom_start=3, min_zoom=1, max_zoom=14)

## Getting All Columns name to show while Hover in any country
columns_popup = happiness_report_mean.columns[2:-2]
# Since there are alot of country it will overcrowded so we user MarkerCluster
marker_cluster = MarkerCluster().add_to(m)

for i in range(166):
    lo = happiness_report_mean.iloc[i, -1]
    la = happiness_report_mean.iloc[i, -2]
    ## Getting Message to Show on Hover
    name = happiness_report_mean.iloc[i, 0]
    mess = "Country" + " : " + str(name)  # First Line Country Name
    
    for i in range(8):
        mess += '<br>'  #Break Line
        mess += str(columns_popup[i]) + " : " +str(np.round(happiness_report_mean[happiness_report_mean['Country name'] == name].values[:,i+2][0],3))
    ## Create Marker over Country
    folium.Marker(location=[la, lo],draggable= True,
                  tooltip= mess,
                  icon=folium.Icon(icon='street-view', prefix='fa'),
                     ).add_to(marker_cluster) ## Addning Our Marker to Cluster
# To add Tiles Layer
folium.TileLayer('OpenStreetMap').add_to(m)
folium.TileLayer('Stamen Toner').add_to(m)
folium.TileLayer('Stamen Water Color').add_to(m)
folium.TileLayer('cartodbpositron').add_to(m)
folium.TileLayer('cartodbdark_matter').add_to(m)
## Adding Button to Change Tiles
folium.LayerControl().add_to(m)

m
```
