```python
# I am reading csv file with pandas library
data = pd.read_csv("../input/videogamesales/vgsales.csv")
```

```python
# I look at the ratio between the information
f,ax = plt.subplots(figsize=(13,13)) # Table size
sns.heatmap(data.corr(), annot = True, linewidths=.5, fmt='.2f',ax=ax) # Annot:transparency Fmt:number of digits after comma
plt.show()
```

```python
# Line Plot
# A Line plot can be defined as a graph that displays data as points or check marks above a number line, showing the frequency of each value
data.NA_Sales.plot(kind = 'line', color = 'g', label = 'NA_Sales', linewidth = 1, alpha = 0.5, grid = True, linestyle = ':')
data.EU_Sales.plot(color = 'r', label = 'EU_Sales', linewidth = 1, alpha = 0.5, grid = True, linestyle = '-.')
plt.legend(loc='upper right')
plt.xlabel('Game')
plt.ylabel('NA_Sales - EU_Sales')
plt.title('Line Plot')
plt.show()
```

```python
# Scatter Plot
# The scatter diagram graphs pairs of numerical data, with one variable on each axis, to look for a relationship between them
data.plot(kind = 'scatter', x = 'Other_Sales', y = 'Global_Sales', alpha = 0.5, color = 'darkblue')
plt.xlabel('Other_Sales')
plt.ylabel('Global_Sales')
plt.title('Attack Defense Scatter Plot')
plt.show()
```

```python
# Histogram
# A histogram is basically used to represent data provided in a form of some groups.It is accurate method for the graphical representation of numerical data distribution
#It is a type of bar plot where X-axis represents the bin ranges while Y-axis gives information about frequency
data.Year.plot(kind = 'hist', bins = 50, figsize = (15,15))
plt.title('Year')
plt.show()
```

```python
# clf() = cleans it up again you can start a fresh
data.Year.plot(kind = 'hist', bins = 50)
plt.show()
##########################
plt.clf()
```
