```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
datum = pd.read_csv("../input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv")
datum.head()
```

```python
datum["stroke"].value_counts().plot(kind = "bar")
plt.title("Stroke");
```

```python
import plotly.express as px
fig = px.histogram(datum, x="age", y="avg_glucose_level", color="work_type", facet_col="work_type",
                  marginal="box")
fig.show();
```

```python
sns.lmplot(x="age", y="avg_glucose_level", hue="gender",
           col="work_type", row="Residence_type", data=datum);
```

```python
sns.pairplot(datum [["age", "avg_glucose_level", "bmi"]], diag_kind="kde");
```

```python
num_datum = pd.Series(datum.columns)

fig = plt.figure(figsize=(8,6))
sns.heatmap(datum[num_datum].cov(),annot=True, fmt=".2f",mask=np.triu(datum[num_datum].cov()),cbar=False)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
Features = ['age','bmi','hypertension','heart_disease','avg_glucose_level']
x = datum[Features]
y = datum["stroke"]
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size= 0.2, random_state= 2)
```

```python
cols_with_missing = [col for col in x_train.columns if x_train[col].isnull().any()]
reduced_x_train = x_train.drop(cols_with_missing, axis = 1)
reduced_x_test = x_test.drop(cols_with_missing, axis = 1)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(reduced_x_train, y_train)
log_reg_pred = log_reg.predict(reduced_x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from mlxtend.plotting import plot_confusion_matrix
cm = confusion_matrix(y_test, log_reg_pred)
plt.figure()
plot_confusion_matrix(cm, figsize=(10,8), hide_ticks=True, cmap=plt.cm.Blues)
plt.title("Logistic Regression Model - Confusion Matrix")
plt.xticks(range(2), ["Yes","No"], fontsize=22)
plt.yticks(range(2), ["Yes","No"], fontsize=22)
plt.show();
```

```python
from sklearn.metrics import classification_report
from sklearn import metrics
print(classification_report(y_test,log_reg_pred))
metrics.plot_roc_curve(log_reg, reduced_x_test, y_test);
```

```python
#Support Vector Machine
from sklearn.svm import SVC
sup_clf = SVC()
sup_clf.fit(reduced_x_train, y_train)
sup_clf_pred = sup_clf.predict(reduced_x_test)
sup_clf_acc = accuracy_score(y_test, sup_clf_pred)

cm = confusion_matrix(y_test, sup_clf_pred)
plt.figure()
plot_confusion_matrix(cm, figsize=(10,8), hide_ticks=True, cmap=plt.cm.Blues)
plt.title("Support Vector Machine - Confusion Matrix")
plt.xticks(range(2), ["Yes","No"], fontsize=22)
plt.yticks(range(2), ["Yes","No"], fontsize=22)
plt.show()
```

```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import metrics
print(classification_report(y_test,sup_clf_pred))
metrics.plot_roc_curve(sup_clf, reduced_x_test, y_test);
```

```python
from sklearn.neighbors import KNeighborsClassifier
#K Neighbours Classifier
kn_clf = KNeighborsClassifier(n_neighbors=6)
kn_clf.fit(reduced_x_train, y_train)
kn_pred = kn_clf.predict(reduced_x_test)
kn_acc = accuracy_score(y_test, kn_pred)

cm = confusion_matrix(y_test, kn_pred)
plt.figure()
plot_confusion_matrix(cm, figsize=(10,8), hide_ticks=True, cmap=plt.cm.Blues)
plt.title("Neighbours Classifier Model - Confusion Matrix")
plt.xticks(range(2), ["Yes","No"], fontsize=22)
plt.yticks(range(2), ["Yes","No"], fontsize=22)
plt.show()
```

```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import metrics
print(classification_report(y_test,kn_pred))
metrics.plot_roc_curve(kn_clf, reduced_x_test, y_test);
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt_clf = DecisionTreeClassifier(max_leaf_nodes=3, random_state=0, criterion='entropy')
dt_clf.fit(reduced_x_train, y_train)
dt_pred = dt_clf.predict(reduced_x_test)
dt_acc = accuracy_score(y_test, dt_pred)

cm = confusion_matrix(y_test, kn_pred)
plt.figure()
plot_confusion_matrix(cm, figsize=(10,8), hide_ticks=True, cmap=plt.cm.Blues)
plt.title("Decision Tree Model - Confusion Matrix")
plt.xticks(range(2), ["Yes","No"], fontsize=22)
plt.yticks(range(2), ["Yes","No"], fontsize=22)
plt.show()
```

```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import metrics
print(classification_report(y_test,dt_pred))
metrics.plot_roc_curve(dt_clf, reduced_x_test, y_test);
```

```python
#Gradient Boosting Model
from sklearn.ensemble import GradientBoostingClassifier
gradientboost_clf = GradientBoostingClassifier(max_depth=2, random_state=1)
gradientboost_clf.fit(reduced_x_train,y_train)
gradientboost_pred = gradientboost_clf.predict(reduced_x_test)
gradientboost_acc = accuracy_score(y_test, gradientboost_pred)

cm = confusion_matrix(y_test, gradientboost_pred)
plt.figure()
plot_confusion_matrix(cm, figsize=(12,8), hide_ticks=True, cmap=plt.cm.Blues)
plt.title("Gredient Boosting Model - Confusion Matrix")
plt.xticks(range(2), ["diabetes","No diabetes"], fontsize=17)
plt.yticks(range(2), ["diabetes","No diabetes"], fontsize=17)
plt.show()
```
