```python
data = pd.read_csv("../input/mushroom-classification/mushrooms.csv")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB
```

```python
le = LabelEncoder()
encoded_series = data[data.columns[:]].apply(le.fit_transform)
```

```python
encoded_series.head()
```

# **train_test_split**

```python
x = encoded_series.iloc[:,1:]
y = encoded_series.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,  x_test ,  y_train , y_test = train_test_split(x,  y , test_size=0.2)
```

# **navie bayes**

```python
gnb = GaussianNB()
y_pred_na = gnb.fit(x_train, y_train).predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_pred_na)
```

# **SVM**

```python
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
y_pred_svm = clf.fit(x_train, y_train).predict(x_test)
```

```python
accuracy_score(y_test, y_pred_svm)
```
