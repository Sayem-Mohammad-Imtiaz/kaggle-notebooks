```python
genre_data = list()
for idx in df.index:
    genre_data+=df.loc[idx, 'Genre']. split(",")  
```

```python
# atsifiltruoju uniklias vertes
# konvertuojant is list i set ismetamos pasikartojan2cis vertes. Todel 
genre_list = set(genre_data)
```

```python
total= []
for genre in genre_list:
    _=filmai.Genre.apply(lambda x: genre in x)
    total.append([genre, _.sum(), 'none_1', 'none_2', 'none_3'])
    #print(genre, _.sum())
total = pd.DataFrame(total,columns = ['Genre','Sum', 'none_1', 'none_2', 'none_3'])
total = total.sort_values(by='Sum',ascending=False)
total
```

```python
total= []
for genre in genre_list:
    _=filmai.Genre.apply(lambda x: genre in x)
    total.append([genre, _.sum()])
    #print(genre, _.sum())
total = pd.DataFrame(total,columns = ['Genre','Sum'])
total = total.sort_values(by='Sum',ascending=False)
total
```

```python
#visualing data using bar chart
plt.figure(figsize=(10,6))
plt.title("Genre populiarity acording to IMDB 1000 popular movies off 2006-2016")
sns.barplot(y=total.Genre, x=total.Sum);
```

```python
# I created a new dataframe based on my previos one and added the most popular genre's as new colums to the new dataframe
# and I used Boolean Indexing to check if the genre was in all the movies.
nf1 = df.copy()
nf1['Action']=nf1.Genre.apply(lambda x: 'Action' in x)
nf1['Drama']=nf1.Genre.apply(lambda x: 'Drama' in x)
nf1['SciFi']=nf1.Genre.apply(lambda x: 'SciFi' in x)
nf1['Horror']=nf1.Genre.apply(lambda x: 'Horror' in x)
nf1['Thriller']=nf1.Genre.apply(lambda x: 'Thriller' in x)
nf1['Comedy']=nf1.Genre.apply(lambda x: 'Comedy' in x)
nf1['Romance']=nf1.Genre.apply(lambda x: 'Romance' in x)
nf1['Adventure']=nf1.Genre.apply(lambda x: 'Adventure' in x)
nf1['Crime']=nf1.Genre.apply(lambda x: 'Crime' in x)
nf1
```

```python
# For further use of this dataframe I converted Boolean Indexing to Integer indexing
nf1[["Action","Drama","SciFi","Horror","Thriller",
     "Comedy","Romance","Adventure","Crime"]] = nf1[["Action","Drama","SciFi","Horror",
                                                     "Thriller","Comedy","Romance","Adventure","Crime"]].astype(int)
```

```python
# I created a new dataFrame with only the column's that I will use further
zanrai=nf1[['Year','Rating','Metascore',"Action","Drama","SciFi","Horror","Thriller","Comedy","Romance","Adventure","Crime"]]
zanrai
```

```python
# Removed all the NaN values
zanrai = zanrai.dropna()
```

How film genre populiarity are changing through the year

```python
genre_list_2
```

```python
df_dict = {}

for genre in genre_list_2:
    df_dict[genre] = zanrai.groupby(['Year'])[genre].agg([sum])
```

```python
# Created 9 new dataframes for individual genre to check how many times they appear every year
action = zanrai.groupby(['Year']).Action.agg([sum])
drama = zanrai.groupby(['Year']).Drama.agg([sum])
sciFi = zanrai.groupby(['Year']).SciFi.agg([sum])
horror = zanrai.groupby(['Year']).Horror.agg([sum])
thriller = zanrai.groupby(['Year']).Thriller.agg([sum])
comedy = zanrai.groupby(['Year']).Comedy.agg([sum])
romance = zanrai.groupby(['Year']).Romance.agg([sum])
adventure = zanrai.groupby(['Year']).Adventure.agg([sum])
crime = zanrai.groupby(['Year']).Crime.agg([sum])
```

```python
crime.head()
```

```python
# Change the name of the column so it would be more easy to read it 
action = action.rename(columns={'sum': 'Action'})
drama = drama.rename(columns={'sum': 'Drama'})
sciFi = sciFi.rename(columns={'sum': 'SciFi'})
horror = horror.rename(columns={'sum': 'Horror'})
thriller = thriller.rename(columns={'sum': 'Thriller'})
comedy = comedy.rename(columns={'sum': 'Comedy'})
romance = romance.rename(columns={'sum': 'Romance'})
adventure = adventure.rename(columns={'sum': 'Adventure'})
crime = crime.rename(columns={'sum': 'Crime'})
```

```python
# I merge all the dataframes
data = pd.merge(action, drama, left_on='Year', right_on='Year')
data1 = pd.merge(data, sciFi, left_on='Year', right_on='Year')
data2 = pd.merge(data1, horror, left_on='Year', right_on='Year')
data3 = pd.merge(data2, thriller, left_on='Year', right_on='Year')
data4 = pd.merge(data3, comedy, left_on='Year', right_on='Year')
data5 = pd.merge(data4, romance, left_on='Year', right_on='Year')
data6 = pd.merge(data5, adventure, left_on='Year', right_on='Year')
data7 = pd.merge(data6, crime, left_on='Year', right_on='Year')
data7
```

```python
# Using line chart I visualize the data
fig, ax = plt.subplots(1, 1, figsize=(14, 6))
plt.title("Popularity of movie genres 2006-2016")
sns.lineplot(data=data7, dashes=False, ax=ax)
ax.set_ylabel("Populiarumas %")

```
