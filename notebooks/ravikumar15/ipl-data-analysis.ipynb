# IPL Data Analysis

**1. Loading Data**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', 100)
ipl = pd.read_csv("../input/ipl-data-set/matches.csv")
```

**2. First 5 Records**

```python
ipl.head(5)
```

**3. Numer of Rows & Columns**

```python
ipl.shape
```

**4. Frequency Of Most Man Of The Match**

```python
ipl['player_of_match'].value_counts().head(1)
```

**5. Top 10 players with most man of the match**

```python
ipl['player_of_match'].value_counts()[0:10]
```

**6. BarPlot of Top 10 players with most man of the match**

```python
ipl['player_of_match'].value_counts()[0:5].keys()
```

```python
plt.figure(figsize=(6,3))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()), list(ipl['player_of_match'].value_counts()[0:5]))
plt.show()
```

**7. Frequeny of result column**

```python
ipl['result'].value_counts()
```

```python
plt.figure(figsize=(6,3))
plt.barh(list(ipl['result'].value_counts().keys()[0:5]), list(ipl['result'].value_counts()), color='green')
plt.show()
```

**8. Frequeny of Toss Winners**

```python
ipl['toss_winner'].value_counts()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(y=ipl['toss_winner'].value_counts().keys(), x=ipl['toss_winner'].value_counts().values,  palette="Accent_r")
plt.title("Maximum toss winners")
plt.show()
```

**9. Records of team who decided to bat first and won**

```python
battingFirst = ipl[ipl['win_by_runs']!=0]
battingFirst.head(5)
```

**Making histogram to show distribution of runs**

```python
plt.figure(figsize=(10,5))
plt.hist(battingFirst['win_by_runs'], color='red')
plt.xlabel("Runs")
plt.ylabel("Number of Matches")
plt.show()
```

**10. Number of wins w.r.t each after after batting first**

```python
battingFirst['winner'].value_counts()
```

**11. Make barplot for top 5 teams with most wins after batting first**

```python
plt.figure(figsize=(13,5))
plt.bar(battingFirst['winner'].value_counts().keys()[0:5],battingFirst['winner'].value_counts()[0:5], color=['Blue','yellow','pink','purple','brown'])
plt.title('Top 5 teams won after batting first')
plt.show()
```

**Make piechart for BATTING first with winning percentage**

```python
plt.figure(figsize=(10,10))
plt.pie(list(battingFirst['winner'].value_counts()),labels=list(battingFirst['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

**12. Records of team who decided to BALL FIRST and won**

```python
ballFirst=ipl[ipl['win_by_wickets']!=0]
ballFirst['winner'].value_counts()
```

Make barplot

```python
plt.figure(figsize=(7,5))
plt.hist(ballFirst['win_by_wickets'], bins=30)
plt.xlabel("Wickets")
plt.ylabel("Number of Matches")
plt.show()
plt.show()
```

**Make barplot for top 5 teams with most wins after balling first**

```python
plt.figure(figsize=(13,5))
plt.bar(ballFirst['winner'].value_counts().keys()[0:5],ballFirst['winner'].value_counts()[0:5], color=['purple','blue','yellow','brown','pink'])
plt.show()
```

**Make piechart for BALLING first with winning percentage**

```python
plt.figure(figsize=(10,10))
plt.pie(list(ballFirst['winner'].value_counts()),labels=list(ballFirst['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
