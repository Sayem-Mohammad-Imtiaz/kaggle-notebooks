**2. LOADING AND EXPLORING DATA**

```python
#Loading the data
data = pd.read_csv('../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')

#First look to the dataset
data.head()
```

**3. MAKING AND ANSWERING QUESTIONS**

**3.1. What is the top 10 best rated books in Amazon?**

```python
data.sort_values('User Rating', ascending=False).head(10)
```

**3.2. What is the top 10 of authors with more books?**

```python
data.groupby('Author')['User Rating'].count().sort_values(ascending=False).head(10)
```

**3.3 Top 10 of best rated authors**

```python
data.groupby('Author')['User Rating'].mean().sort_values(ascending=False).head(10)
```

**3.4. Bestsellers' ratings distribution**

```python
sns.distplot(a=data['User Rating'], kde=False)
```

**3.5. Relation rate/price**

```python
sns.regplot(x=data['Price'], y=data['User Rating'])
```

User Ratings show a slight negative correlation with Price of the Bestselling Book but as the 90% of prices are under 40$, that is not a conclusion we can take.
