```python
df=data[data["Datum"].apply(lambda x: x[-3:])=="UTC"]
```

```python
data["Luan_Country"]=data["Location"].apply((lambda x:x.split(",")[-1]))
```

### Lets See the Success Rate in Top 10 Companies and Top 10 countries

```python
data["Status Mission"]=np.where(data["Status Mission"]=="Success",1,0)
```

```python
top_com=list(data["Company Name"].value_counts().head(11).index)
```

```python
data[data["Company Name"].isin(top_com)].groupby("Company Name")["Status Mission"].mean().sort_values(0,ascending=False,).iplot(kind="bar",theme="solar",title="Success rate in Each company")
```

```python
top_count=list(data["Luan_Country"].value_counts().head(11).index)
```

```python
data[data["Luan_Country"].isin(top_count)].groupby("Luan_Country")["Status Mission"].mean().sort_values(0,ascending=False,).iplot(kind="bar",theme="solar",title="Success rate in Each Country")
```
