```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
HeartData['gender'] = le.fit_transform(HeartData['gender'])
HeartData['currentSmoker'] = le.fit_transform(HeartData['currentSmoker'])
HeartData['BPMeds'] = le.fit_transform(HeartData['BPMeds'])
HeartData['prevalentStroke'] = le.fit_transform(HeartData['prevalentStroke'])
HeartData['prevalentHyp'] = le.fit_transform(HeartData['prevalentHyp'])
HeartData['diabetes'] = le.fit_transform(HeartData['diabetes'])
```

```python
X = HeartData.drop('TenYearCHD',axis=1)
y = HeartData['TenYearCHD']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(max_iter=10000)
```

```python
logreg.fit(X_train,y_train)
```

```python
logreg.score(X_train,y_train)
```

```python
y_pred = logreg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred)*100)
```
