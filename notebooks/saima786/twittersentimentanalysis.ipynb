# Importing all the libraries required

```python
import pandas as pd
```

```python
from nltk.tokenize import word_tokenize
from sklearn import linear_model
from sklearn import metrics
from sklearn import model_selection
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

# Read train and test files

```python
train_df=pd.read_csv("/kaggle/input/twitter-sentiment-analysis-hatred-speech/train.csv",error_bad_lines=False)
```

```python
test_df=pd.read_csv("/kaggle/input/twitter-sentiment-analysis-hatred-speech/test.csv",error_bad_lines=False)
```

```python
count_vec=CountVectorizer(tokenizer=word_tokenize,token_pattern=None)
```

```python
count_vec.fit(train_df.tweet)
```

```python
xtrain=count_vec.transform(train_df.tweet)
xtest=count_vec.transform(test_df.tweet)
```

# Logistic Regression Model

```python
model=linear_model.LogisticRegression()
model.fit(xtrain,train_df.label)
train_pred=model.predict(xtrain)
accuracy=metrics.accuracy_score(train_df.label,train_pred)
test_pred=model.predict(xtest)
```

```python
accuracy
```

```python
submission = pd.DataFrame(test_pred)
submission.to_csv("submission.csv", header=True)    
```
