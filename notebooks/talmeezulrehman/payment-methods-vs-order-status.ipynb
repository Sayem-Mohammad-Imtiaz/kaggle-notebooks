```python
data = pd.read_csv("../input/pakistans-largest-ecommerce-dataset/Pakistan Largest Ecommerce Dataset.csv")
data.head()
```

```python
data.shape
```

```python
#Drop Column Only if All the Values are Null
data = data.dropna(how='all', axis=1)
```

```python
data.shape
```

```python
data.head()
```

```python
#Drop Row Only if All the Values are Null
data = data.dropna(how='all', axis=0)
data.shape
```

```python
pie_status=data['status'].value_counts()
pie_status
```

```python
data['payment_method'].unique()
```

```python
data['payment_method'].value_counts()
```

```python
df_compl = data[(data['status'] == 'complete') | (data['status'] == 'paid') | (data['status'] == 'received') | (data['status'] == 'closed') | (data['status'] == 'cod')]
df_compl.tail(10)
```

```python
import plotly.express as px
pay_methd = df_compl['payment_method'].value_counts().nlargest(10).to_frame()
px.bar(pay_methd, y = pay_methd['payment_method'], x = pay_methd.index, title = 'Mostly used payment method for completed orders')
```

```python
df_refund = data[(data['status'] == 'order_refunded') | (data['status'] == 'exchange') | (data['status']== 'refund')]
df_refund.head()
```

```python
pay_methd = df_refund['payment_method'].value_counts().nlargest(10).to_frame()
px.bar(pay_methd, y = pay_methd['payment_method'], x = pay_methd.index, title = 'Mostly used payment method for refund orders')
```

```python
df_cancel = data[(data['status'] == 'canceled')]
df_cancel.head()
```

```python
pay_methd = df_cancel['payment_method'].value_counts().nlargest(10).to_frame()
px.bar(pay_methd, y = pay_methd['payment_method'], x = pay_methd.index, title = 'Mostly used payment method for canceled oredrs')
```

```python
data.isnull().sum()
```

```python
df1 = data.groupby(['FY','payment_method']).size().reset_index(name='count')
df2 = data.groupby(['FY','payment_method'])['grand_total'].sum().reset_index(name='sum')
```

```python
df1
```

```python
df2
```

```python
temp = pd.concat([df1, df2['sum']], axis=1)
temp
```

```python
fig = px.bar(temp, x="FY", y="count", color="payment_method", title="Yearly Transactions by Payment method")
fig.show()
```
