# ***But we need only those categories in which order status is completed because we want to find the BEST SELLING CATEGORY***

```python
data['status'].unique()
```

```python
df1 = data[(data['status'] == 'complete') | (data['status'] == 'paid') | (data['status'] == 'received')]
df1.head(10)
```

 **this is the total numbers completed orders**

```python
df1.shape[0] 
```

# Visualization of Top 10 selled category

```python
import plotly.express as px
catg_chart = df1['category_name_1'].value_counts().nlargest(10).to_frame()
px.bar(catg_chart, y = catg_chart['category_name_1'], x = catg_chart.index, title = 'Most Selling category')
```
