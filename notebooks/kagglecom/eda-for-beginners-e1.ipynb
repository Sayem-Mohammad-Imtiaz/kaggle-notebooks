```python
# This Python 3 environment 
#import data

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read dataset
df = pd.read_csv('../input/montcoalert/911.csv')
```

```python
#read three top head from dataset
df.head(3)
```

```python
# Finding information from data frame
df.info()
```

```python
#Top 3 zipcodes for 911 calls
df['zip'].value_counts().head(3)
```

```python
#Top 3 timeStamp for 911 calls
df['timeStamp'].value_counts().head(3)
```

```python
#The number of unique title codes 
df['title'].nunique()
```

```python
# Using pd.to_datetime to convert Timestamps from strings to  objects
df['timeStamp']=pd.to_datetime(df['timeStamp'])

# Then create day of week column by .apply
df['day of week']=df['timeStamp'].apply(lambda x:x.dayofweek)
df.head()
```

```python
#Using  dictionary and .map() to map the string names to defined day of the week
week_dictionary={0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['day of week']=df['day of week'].map(week_dictionary)
df.head()
```

```python
#Using seaborn to create a countplot based on Day of Week and Town
fig,axes=plt.subplots(figsize=(20,12))
sns.countplot(data=df, x='day of week', hue='twp')
```

```python
# create month column
df['month']=df['timeStamp'].apply(lambda x:x.month)
df.head()
```

```python
#Using seaborn to create a countplot based on month and Town
fig,axes=plt.subplots(figsize=(18,10))
sns.countplot(data=df, x='month', hue='twp')
```

```python
#Creating object by .groupby to fill some information 
byMonth = df.groupby('month').count()
byMonth
```

```python
# creating plot based on numberof towns and month
byMonth['twp'].plot()
```

```python
#create reason
df['Reason']=df['title'].apply(lambda x: x.split(':')[0])
df.head()

fig,axes=plt.subplots(figsize=(18,8))
df[df['Reason']=='Traffic'].groupby('month').count()['Reason'].plot(color='r',fontsize=22)
plt.title('Traffic',fontsize=22, color='b')
plt.xlabel('month',fontsize=22,color='b')
```
