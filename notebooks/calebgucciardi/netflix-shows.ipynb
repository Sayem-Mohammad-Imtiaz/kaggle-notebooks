```python
df = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')
df.sample(3)
```

```python
len(df)
```

```python
movie_filter = df['type'] == 'Movie' # creare filtri da utilizzare come colonne # questo seleziona True solo per i film
movie_filter
```

```python
just_movies = df[movie_filter]
just_movies.sample(5)
```

```python
just_movies['release_year'].describe() # media 2012 - mediana 2016
```

```python
year_filter = just_movies['release_year']>=2000
genz_movies = just_movies[year_filter]
genz_movies['release_year'].hist(bins=10)
```

```python
# Grafico a torta degli anni
count_movies_per_year = genz_movies.groupby(by='release_year').agg('count')['show_id']
count_movies_per_year.plot.pie(y='release_year',figsize=(8,8))
```

```python
category = []
for s in genz_movies['listed_in']:
    for i in s.split(', '):
        category.append(i)

data = []
for i in set(category):
    data.append(category.count(i))
count_movies_per_category = pd.DataFrame({'c':data},index=set(category))
count_movies_per_category.plot(kind='barh',figsize=(8,8))
```
