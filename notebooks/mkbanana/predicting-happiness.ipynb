Hey. I am a new machine learning student and i am simply using this dataset to get the hang of it.
I used the "2015 world-happiness" dataset.

```python
df2015 = pd.read_csv('/kaggle/input/world-happiness/2015.csv')

#Presense of missing values
missing = df2015.isnull().sum().sort_values(ascending=False)
print(missing)
```

Awesome! we have no missing values in our dataset.

Making a correlation matrix.
I create one to enable me to see which features have more weight in our happiness score. If i am using it wrong you can inform me in the comments.

```python
import seaborn as sns
cor_matrix = df2015.corr()
k = 10 # the number off variables in the matrix
cols = cor_matrix.nlargest(k, 'Happiness Score')['Happiness Score'].index
cm = np.corrcoef(df2015[cols].values.T)
sns.set(font_scale=1)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 8}, yticklabels=cols.values, xticklabels=cols.values)
```

The Gdp, Family and health have a higher percentage hence, they are features that are a must for our predictions. However despite freedom and Dystopia residual having less weight than the three they are still an important feature for better predictions.

```python
features = ['Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom', 'Dystopia Residual']
x = df2015[features]
y = df2015['Happiness Score']
```

With our features in place. Time to split the data into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=1)
```

The data has been split.Lets take a look at their shape(number of rows and columns).


```python
print(x_train.shape)
print(x_test.shape)
```

Time to fit the data in a classifier. I had looked at various regression models and HuberRegressor gave the highest score.

```python
from sklearn.linear_model import HuberRegressor
clf = HuberRegressor()
clf.fit(x_train, y_train) # training the model
```

With our model trained next is to test the model and get its accuracy.

```python
clf.predict(x_test)
my_score = clf.score(x_test, y_test)
print(my_score)
```

98% accuracy. 2% off but not bad, still usable.With the model done. I got curious and wondered about the GDP over the years and made a small graph to see them.
