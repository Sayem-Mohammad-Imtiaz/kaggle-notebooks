## Quick see to data

```python
data = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')
data.head()
```

## Let's look up to series and film count

```python
sns.countplot(x='type',data=data)
plt.show()
```

```python
fig = px.pie(values=data['type'].value_counts(),names=data['type'].value_counts().index,title='Count of Type')
fig.show()
```

## In what year was the contents published?

```python
netflix_date = data[['date_added']].dropna()
netflix_date['year'] = netflix_date['date_added'].apply(lambda x : x.split(', ')[-1]).astype(int).sort_values(ascending=True)
plt.figure(figsize=(12,10))
sns.countplot(x='year',data=netflix_date)
plt.show()
```

## What kind have content in Netflix ?

```python
data['listed_in'] = data['listed_in'].dropna()
category= pd.DataFrame(columns=['category'])
for i in data['listed_in']:
    for j in i.split(', '):
        category=category.append({'category':j},ignore_index=True)
```

```python
fig = px.pie(values=category['category'].value_counts(), names=category['category'].value_counts().index,
             title='What kind have content in Netflix',labels={'label':'Deneme'},height=600)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

## How many country published content in Netflix ?

```python
data = data.dropna()
country= pd.DataFrame(columns=['country'])
for i in data['country']:
    for j in i.split(', '):
        country=country.append({'country':j},ignore_index=True)
```

```python
print(str(len(country['country'].unique()))+" country was published content in Netflix")
```

## Which 10 country published more content in Netflix?

```python
most_country = country['country'].value_counts().iloc[:10]
fig = px.histogram(x=most_country.index,y=most_country.values,labels={'x':'Country','y':'Content'})
fig.show()
```
