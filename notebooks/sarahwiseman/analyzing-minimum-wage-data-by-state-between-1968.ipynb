# Analyzing Minimum Wage Data by State between 1968 and 2020

```python
import numpy as np
import pandas as pd
data = pd.read_csv('../input/us-minimum-wage-by-state-from-1968-to-2017/Minimum Wage Data.csv', encoding = 'latin')
data.head()
```

Reading in the data

Setting Year and State Indexes

```python
df = data.set_index(['Year','State'])
df[0:60]
```

Checking for any null values before copmutation

```python
df.notnull().any()
```

Finding the states with a Minimum Wage of $0.00 in 2020.

```python
df2 = df[df['State.Minimum.Wage']==0]
df22 = df2.query('Year == 2020')
df22.index
```

Finding the Federal Minimum Wage by Year

```python
df['Federal.Minimum.Wage'].mean(level = ['Year'])
```

Counting the number of years each state's Minimum Wage was below the Federal Minimum Wage

```python
df4 = df[df['State.Minimum.Wage']<df['Federal.Minimum.Wage']]
s1 = df4['State.Minimum.Wage'].notna().sum(level = ['State'])
s1.sort_values()
```

Counting the number of years each state's Minimum Wage was equal to its Effective Minimum Wage

```python
df6 = df[df['State.Minimum.Wage']==df['Effective.Minimum.Wage']]
s2 = df6['State.Minimum.Wage'].notna().sum(level = ['State'])
s2.sort_values()
```

Calculating the inflation between each year and 2020

```python
df3 = (df['State.Minimum.Wage.2020.Dollars']-df['State.Minimum.Wage'])/df['State.Minimum.Wage']
df3.mean(level = 'Year')
```
