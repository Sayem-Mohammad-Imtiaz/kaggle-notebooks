```python
df_orders=pd.read_csv("/kaggle/input/city-cleaned-by-books/GP_Orders_5_excel_fuzzy_6b.csv")
```

```python
df_orders.head()
```

```python
df_orders = df_orders.astype(str).apply(lambda x: x.str.lower())
df_orders.rename({'Column1':'city','Total items':'total_items','Order Number':'order_number','Order Status':'order_status','Book Name':'book_name','Order Date & Time':'order_date_time','City':'old_city','Payment Method':'payment_method','Total weight (grams)':'total_weight'}, axis=1, inplace=True)
```

```python
# Replace cash on delivery (cod) as cash on delivery

df_orders['payment_method'].replace('cash on delivery (cod)','cash on delivery', inplace=True)

#Replace all payment methos of non as cash
df_orders['payment_method'].replace('nan','cash', inplace=True)

#Convert Order Status, Total items and Total weight to float and Order Date & Time as DATE format
df_orders[['total_items','total_weight','total_items_record']]=df_orders[['total_items','total_weight','total_items_record']].astype('float')

# convert Order Number to integer
df_orders[['order_number']]=df_orders[['order_number']].astype('int')
```

```python
df_orders.head(5)
```

```python
df_orders=df_orders.sort_values('total_items',ascending=False)
```

```python
df_orders_status=df_orders['order_status'].unique().tolist()
df_orders_status
```

```python
################  NUMBERS OF BOOKS BY CITY FOR COMPLETED RETURNED AND CANCLLED ###########################
```

```python
for x in range(0,3):
    df_orders_x=df_orders[df_orders['order_status']== df_orders_status[x]]
    df_orders_x_city=df_orders_x.groupby(['city'])[['order_number']].count()
    df_orders_x_city.sort_values('order_number', ascending=False)
    df_orders_x_city=df_orders_x_city.sort_values('order_number', ascending=False).head(5)
    
    if df_orders_status[x]=='returned':
        df_orders_returned_city=df_orders_x_city
#         print(df_orders_status[x])
#         print(df_orders_returned_city)
#         print(' ')
#         df_orders_returned_city.plot.pie(title ='Returned',autopct='%1.1f%%',subplots=True,legend=False)
#         df_orders_returned_city.plot.bar(title='Returned',subplots=True,figsize=(5,5))
       
        
    elif df_orders_status[x]=='completed':
        df_orders_completed_city=df_orders_x_city
#         print(df_orders_status[x])
#         print(df_orders_completed_city)
#         print(' ')
#         df_orders_completed_city.plot.pie(title ='COMPLETED',autopct='%1.1f%%',subplots=True,legend=False)
#         df_orders_completed_city.plot.bar(title ='COMPLETED',subplots=True,figsize=(5,5))
       
        
    elif df_orders_status[x]=='cancelled':
        df_orders_cancelled_city=df_orders_x_city 
#         print(df_orders_status[x])
#         print(df_orders_cancelled_city)
#         print(' ')
#         df_orders_cancelled_city.plot.pie(title ='CANCELLED',autopct='%1.1f%%',subplots=True,legend=False)
#         df_orders_cancelled_city.plot.bar(subplots=True,figsize=(5,5))
    

        
    #else print('SEE LIST ERROR'):
    #print(df_orders_status[x])
   
    #print(df_orders_returned_city)
    #print(' ')
```

```python
cmp_x = df_orders_completed_city.plot(title='Completed',kind='bar')
ren_x = df_orders_returned_city.plot(title='Returned',kind='bar', color='red')
can_x = df_orders_cancelled_city.plot(title='Cancelled',kind='bar', color='green')
```
