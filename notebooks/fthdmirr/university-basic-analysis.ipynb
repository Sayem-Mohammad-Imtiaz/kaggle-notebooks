```python
#Let's say university research 60 and above Class 1,40-60  Class 2, 0-40 Class 3 


uni_data['university_class']=[1 if i>=60  else 2  if 60>i >=40 else 3 for i in uni_data['research']]
```

```python
def bar_plot(variable):
    """
    input:variable (example:university class)
    output: bar plot & value count
    """
    #get feature
    var=uni_data[variable]
    # count number of categorical variable(value/sample)
    varValue=var.value_counts()
    #visualize
    plt.figure(figsize=(10,4))
    plt.bar(varValue.index,varValue)
    plt.xticks(varValue.index,varValue.index.values)
    plt.ylabel("Frequence")
    plt.title(variable)
    plt.show()
    print('{} : \n {}'.format(variable,varValue))
```

```python
category=['university_class']
for x in category:
    bar_plot(x)
```

## Outlier Detection

```python
def detect_outliers(df,features):
    outlier_indices=[]
    for c in features:
        #1st quarties
        Q1=np.percentile(df[c],25)
        #3rd quarties
        Q3=np.percentile(df[c],75)
        #IQR
        IQR=Q3-Q1
        #Outlier step
        outlier_step=IQR*1.5
        #detect outlier and their indeces
        outlier_list_col=df[(df[c]<Q1 - outlier_step) | (df[c]> Q3+outlier_step)].index
        #store indeces
        outlier_indices.extend(outlier_list_col)
    outlier_indices=Counter(outlier_indices)
    multiple_outliers=list(i for i, v in outlier_indices.items() if v>2)
    return multiple_outliers
    
```
