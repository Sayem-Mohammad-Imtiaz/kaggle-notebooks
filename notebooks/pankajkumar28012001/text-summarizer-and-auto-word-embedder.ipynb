```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
a=spacy.load("en_core_web_sm")
d=a(text)
w_freq={}
s_w=list(STOP_WORDS)
for w in d:
    if w.text.lower() not in s_w:
        if w.text.lower() not in punctuation:
            if w.text not in w_freq.keys():
                w_freq[w.text]=1
            else:
                w_freq[w.text]+=1           
```

```python
s_t=[s for s in d.sents]
s_score={}
for s in s_t:
    for w in s:
        if w.text.lower() in w_freq.keys():
            if s not in s_score.keys():
                s_score[s]=w_freq[w.text.lower()]
            else:   
                s_score[s]+=w_freq[w.text.lower()]
```

```python
from heapq import nlargest
summ=nlargest(5,iterable=s_score,key=s_score.get)
print(summ)
```
