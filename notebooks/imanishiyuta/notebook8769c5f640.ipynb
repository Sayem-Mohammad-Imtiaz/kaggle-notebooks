```python
data = pd.read_csv('../input/housesalesprediction/kc_house_data.csv')
```

```python
data['date']=data['date'].replace('\S000000', '', regex=True)
data['bathrooms']=(data['bathrooms']*100).astype(int)
data['lat']=(data['lat']*10000).astype(int)
data['long']=(data['long']*1000).astype(int)
data['floors']=data['floors'].astype(int)
```

```python
data
```

```python
x = data.drop(['id','price'], axis=1)
y = data['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
# 決定木分析
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier 
from sklearn.tree import plot_tree

# グラフの大きさを調整する
plt.rcParams["figure.figsize"] = (20, 20)
clf = DecisionTreeClassifier(random_state=0, max_depth=3)
clf = clf.fit(x_train, y_train)
plot_tree(
    clf,
    feature_names=x.columns,
    filled=True, rounded=True
)

plt.show()
plt.savefig("tree1.png")
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(penalty='l2', solver="sag", random_state=0)
```

```python
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```
