# **READING THE DATA SET 

```python
import pandas as pd
df= pd.read_csv("/kaggle/input/indian-food-101/indian_food.csv")
df1= df[['name','diet']]
df1
```

# **TOTAL DISHES ON THE BASIS OF DIET**

```python
df2 = df1.groupby('diet').count();
df2.columns=['count']
df2
```

# **Percentage of Vegetarian and Non Vegetarian Dishes**

```python
df2['percent'] = (df2['count'] / df2['count'].sum()) * 100
df2
```

# **IN THE PROVIDED DATA SET THERE IS MORE VEGETARIAN DISHES AND COMPARED TO NON VEGETARIAN DISHES  **

```python
plot = df2.plot.pie(y='percent', figsize=(5, 5))
```

 # COMPARING DISHES ON THE BASIS OF DIET AND COOKING TIME THEY TOOK **

```python
df3=df[['name','cook_time','diet']]
df4=df3.groupby('diet').mean()
df4
```

# **COMPARING DISHES ON THE BASIS OF DIET AND PREPARATION TIME THEY TOOK**

```python
df3=df[['name','prep_time','diet']]
df5=df3.groupby('diet').mean()
df5
```

# **AVERAGE TIME TAKEN BY BOTH THE DISHES IN THIER COOKING AND PREPARATION PERIOD**

```python
df6=[df4,df5]
result=pd.concat(df6,axis=1)
result
```

# **COMPARING THE TIME TAKEN BY BOTH THE DISHES IN THIER COOKING AND PREPARATION PERIOD WITH THE HELP OF A GRAPH**

```python
plot = result.plot.pie(subplots=True, figsize=(11, 6))
```
