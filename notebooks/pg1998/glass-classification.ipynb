```python
import seaborn as sns
import matplotlib.pyplot as plt

corr = data.corr()
#Plot figsize
fig, ax = plt.subplots(figsize=(10, 8))
#Generate Heat Map, allow annotations and place floats in map
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt=".2f")
#Apply xticks
plt.xticks(range(len(corr.columns)), corr.columns);
#Apply yticks
plt.yticks(range(len(corr.columns)), corr.columns)
#show plot
plt.show()
```

```python
x=data.iloc[:,0:9]
y=data.iloc[:,9]
print(x.head())
```

```python
print(y.head())
```

```python
x.describe()
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 42)
print(X_train.shape)
print(X_test.shape)
```

```python

dct = DecisionTreeClassifier(class_weight='balanced')
dct_model = dct.fit(X_train,y_train)
y_pred_dct = dct_model.predict(X_test)
con_fu_dct = confusion_matrix(y_test,y_pred_dct)
clfreport_dct = classification_report(y_test,y_pred_dct)

```

```python
sns.heatmap(data = con_fu_dct,annot = True,fmt = "d",cmap="YlGnBu")
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix for Decision Tree Classifier')
```

```python
print(clfreport_dct)
```

```python
rft = RandomForestClassifier(n_estimators = 150,random_state = 42,class_weight='balanced')
rft_model = rft.fit(X_train,y_train)
y_pred_rft = rft_model.predict(X_test)
con_fu_rft = confusion_matrix(y_test,y_pred_rft)
clfreport_rft = classification_report(y_test,y_pred_rft)
```

```python
sns.heatmap(data = con_fu_rft,annot = True,fmt = "d",cmap="YlGnBu")
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix for Random Forest Classifier')

```

```python
print(clfreport_rft)
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax = MinMaxScaler()
X_s = minmax.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
X_train1,X_test1,y_train1,y_test1 = train_test_split(X_s,y,test_size = 0.3,random_state = 42)
print(X_train1.shape)
print(X_test1.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logit_reg = LogisticRegression(random_state=42,class_weight='balanced',solver='newton-cg',multi_class='multinomial')
logit_reg_model = logit_reg.fit(X_train1,y_train1)
y_pred_logit = logit_reg_model.predict(X_test1)
cf_logit = confusion_matrix(y_test1,y_pred_logit)
clss_report_logit = classification_report(y_test1,y_pred_logit)
```

```python
sns.heatmap(data = cf_logit,annot = True,fmt = "d",cmap="YlGnBu")
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix for Logistic regression')
```

```python
print("Classification report : ")
print(clss_report_logit)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb_model=gnb.fit(X_train1,y_train1)
y_pred_gnb=gnb_model.predict(X_test1)
con_fu_gnb = confusion_matrix(y_test1,y_pred_gnb)
clfreport_gnb = classification_report(y_test,y_pred_gnb)

```

```python
sns.heatmap(data = con_fu_gnb,annot= True,fmt = "d",cmap="YlGnBu")
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix for Naive Bayes')
```

```python
print(clfreport_gnb)
```

**Using PCA**

```python
from sklearn.decomposition import PCA
pca = PCA()
X_s_pca = pca.fit(X_s)
```

```python
plt.figure()
plt.plot(np.cumsum(X_s_pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Glass Classification Dataset Explained Variance')
plt.show()
```

```python
pca = PCA(n_components=7)
X_s_pca = pca.fit_transform(X_s)
```

```python
X_train_pca,X_test_pca,y_train_pca,y_test_pca = train_test_split(X_s_pca,y,test_size = 0.3,random_state = 42)
print(X_train_pca.shape)
print(X_test_pca.shape)
```

```python

dct = DecisionTreeClassifier()
dct_model = dct.fit(X_train_pca,y_train_pca)
y_pred_dct = dct_model.predict(X_test_pca)
con_fu_dct = confusion_matrix(y_test_pca,y_pred_dct)
clfreport_dct = classification_report(y_test_pca,y_pred_dct)
```

```python
sns.heatmap(data = con_fu_dct,annot = True,fmt = "d",cmap="YlGnBu")
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix for Decision Tree Classifier')
```

```python
print(clfreport_dct)
```

```python
rft = RandomForestClassifier(n_estimators = 100,random_state = 42,class_weight='balanced')
rft_model = rft.fit(X_train_pca,y_train_pca)
y_pred_rft = rft_model.predict(X_test_pca)
con_fu_rft = confusion_matrix(y_test_pca,y_pred_rft)
clfreport_rft = classification_report(y_test_pca,y_pred_rft)
```

```python
sns.heatmap(data = con_fu_rft,annot = True,fmt = "d",cmap="YlGnBu")
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.title('Confusion Matrix for Random Forest Classifier')

```

```python
print(clfreport_rft)
```
