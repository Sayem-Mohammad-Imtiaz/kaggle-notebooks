# * NESTED FUNCTION

```python
#nested function
def square():     #bir vakue'nin karesini alir. 
    
    def add():
      
        x = 2
        y = 3
        z = x + y
        return z
    return add()**2   #add hesaplandıktan sonra ''**2'' karesini alir print eder.
print(square())    
```

# *  DEFAULT and FLEXIBLE ARGUMENTS

```python
# default arguments
def f(a, b = 1, c = 2):
    y = a + b + c
    return y
print(f(5))      # f degeri 5 olursa yalnızca a degerini 5 yapar. y toplandigi zaman sonuc 5+1+2=8 seklinde olur.
# 
print(f(5,4,3))  # y degerleri 5,4,3 olan degerleri toplanir.
```

```python
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4)

# flexible arguments 
# **kwargs 

def f(**kwargs):
    
    for key, value in kwargs.items():              
        print(key, " ", value)
f(country = 'spain', capital = 'madrid', population = 123456) 
# f degerleri fonksiyon icerisine dictionary seklinde gonderilir. **kwargs komutu bunu saglamaya yarar.
```

# *  LAMBDA FUNCTION

```python
# lambda function

def square(x) :
    return x**2     # karesini alir

print(square(5))

square = lambda x: x**2     #karesini alir.
print(square(4))

tot = lambda x,y,z: x+y+z   # lambda x,y,z degerlerini toplar. 
print(tot(1,2,3))
```

```python
# Example of list comprehension
num1 = [1,2,3]
num2 = [i + 1 for i in num1 ]
print(num2)

#i'yi dongu icerisinde dondur ve 1 artirarak yazdir.
```

```python
# Conditionals on iterable

num1 = [5,10,15]

num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]

print(num2)
# num1 i dongu icerisine al i=10 ise karesini al,eger i <7 ise i-5 degerini al, hicbiri degilse i+5 yap.
```
