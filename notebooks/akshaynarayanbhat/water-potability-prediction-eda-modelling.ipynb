```python
def Pie_chart(data, column_name):
    
    '''
    Function to create a pie chart;
        data: takes the data as input.
        column_name: takes the column name as input.
    '''
    
    values = data[column_name].value_counts() # taking the values of a particular columns
    labels = data[column_name].unique() # taking the unique data
    pie, ax = plt.subplots(figsize = (12,6))
    
    patches, text, autotext = ax.pie(values, labels = labels, shadow = True, autopct = '%1.2f%%', pctdistance = 0.5, explode = [0.06]*data[column_name].unique())
    
    plt.title(column_name, color='white') #title of the chart
    plt.legend(patches, labels, loc='best') # display the legend
    plt.setp(text, color='black') # text color
    plt.axis('equal')
    plt.setp(autotext, color='black')# text color
    autotext[0].set_color('black')
    autotext[1].set_color('red')
    plt.tight_layout()
    plt.show()
    
Pie_chart(data, 'Potability') # function calling
```

```python
def correct_skewness(data):
    
    '''
        Function to correct the data skewness;
            data: takes data as input.
                (Funtion takes data for the operation and uses except column to exclude that particular colum
                from the datset.)
    '''
    data_set = prep.skewcorrect(data, except_columns=['Potability'])
    return data_set

data_set = correct_skewness(data) # function calling
```

```python
X = data_set.iloc[:,:-1].values
y = data_set.iloc[:,-1].values

X.shape, y.shape
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0) # data splitting 80-20% to avoid over fitting
```

```python
X_train, y_train = shuffle(X_train, y_train) # data shuffling
```

```python
# pipelines

pipeline = make_pipeline(RobustScaler()) # creating a pipeline for all the models
Random_forest = make_pipeline(pipeline, RandomForestClassifier(random_state=0, min_samples_leaf = 2, n_estimators = 1000))
Decision_tree = make_pipeline(pipeline, DecisionTreeClassifier(random_state=0))
Logistic_regression = make_pipeline(pipeline, LogisticRegression(random_state=0))
svc = make_pipeline(pipeline, SVC(random_state=0))
KNeighbors = make_pipeline(pipeline, KNeighborsClassifier())
Ada_boost = make_pipeline(pipeline, AdaBoostClassifier(random_state=0))
xgboost = make_pipeline(pipeline, XGBClassifier())
gradientboost = make_pipeline(pipeline, GradientBoostingClassifier(random_state=0))
```

```python
param_dist = {
    'RandomForest':Random_forest,
    'DecisionTree':Decision_tree,
    'LogisticRegression':Logistic_regression,
    'svc':svc,
    'KNeighbors':KNeighbors,
    'AdaBoost':Ada_boost,
    'XGB':xgboost,
    'GD':gradientboost
}
```

```python
def MODEL(model):
    
    '''
    The function to create a model with accuracy score and confusion matrix;
        model: Takes different model alogirthm as an input for the operation.
    '''
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print('The accuracy score of the model is: {}%'.format(accuracy_score(y_test, y_pred)))
    print('-'*50)
    print(confusion_matrix(y_test, y_pred))
    print(classification_report(y_test, y_pred))
```

```python
def model_evaluation(parameter_dictionary):
    
    '''
        Function to find the model accuracy score, classification report;
            parameter_dictionary: takes as an input, which is a dictionary containing the pipeline for the
                                  individual model.
    '''
    
    for name, model in parameter_dictionary.items():
        print('-'*50)
        print(name)
        evaluation = MODEL(model)
    return evaluation
evaluation = model_evaluation(param_dist)
```

## **Let's create Artifical Neural Network model. Let's see whether this model can bring better accuracy scores with greater recall and precision.**

```python
# Model Architecture

model = tf.keras.models.Sequential()

model.add(tf.keras.layers.Dense(units = 16, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 32, activation = 'relu'))
model.add(tf.keras.layers.Dropout(0.3))

model.add(tf.keras.layers.Dense(units = 64, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 128, activation = 'relu'))
model.add(tf.keras.layers.Dropout(0.25))

model.add(tf.keras.layers.Dense(units = 252, activation = 'relu'))
model.add(tf.keras.layers.Dense(units = 516, activation = 'relu'))
model.add(tf.keras.layers.Dropout(0.2))

# output layer
model.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid')) #output for potable or non potable
```

```python
model.compile(optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])
```

```python
M = model.fit(X_train, y_train, batch_size=12, epochs=100, verbose=1)
```

```python
model_evaluation = model.evaluate(X_test, y_test)
```

```python
plt.plot(M.history['accuracy'], 'bo-', label = 'accuracy')
plt.title('Model Traning Accuracy', weight='bold')
plt.legend();
```

```python
plt.plot(M.history['loss'], label = 'loss', color='red')
plt.title('Model Traning Loss', weight='bold')
plt.legend();
```
