```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as ws
ws.filterwarnings("ignore")
```

```python
df = pd.read_csv("/kaggle/input/us-police-shootings/shootings.csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
sns.set()
sns.countplot(df["gender"])
plt.title("The Gender of Killed Person")
plt.show()
```

```python
avg_age_of_person = df.age.mean()
print(avg_age_of_person)
```

```python
# status of arming 
temp = df.armed.value_counts().sort_values(ascending = False).reset_index()
```

```python
sns.barplot(x="index", y = "armed", data = temp[temp["armed"] > 50])
plt.show()
```

Most of them were armed by the gun 

```python
plt.figure(figsize= (20, 20))
sns.barplot(y="index", x = "state", data =  df.state.value_counts().reset_index())
plt.title("State wise count for the killing criminals")
plt.show()
```

```python
sns.countplot(df.signs_of_mental_illness)
plt.title("Status of mental illness")
plt.show()
```

```python
sns.barplot(x="arms_category" , y ="index", data = df.arms_category.value_counts().reset_index())
plt.show()
```
