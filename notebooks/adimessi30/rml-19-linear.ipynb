# Selecting the relevant independent variables

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style(style='ticks')
sns.pairplot(df)
```

# Primary assumption of linear regression is a linear relationship exists between the dependent and independent variable. From the above pairplot, the following independent variables seem to have linear relationship with sqft_living(target variable):
1. Price
2. Bathrooms
3. Grade
4. Sqft_above
5. Sqft_basement
6. Sqft_living15

```python
# Filtering down only the required independent variables
subset_df = df[['sqft_living', 'price', 'bathrooms', 'grade',
       'sqft_above', 'sqft_basement', 'sqft_living15']]
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(subset_df.corr().abs(), annot=True)
```

# The correlation matrix confirms the linear relationship between the filtered independent variables and the target variable but for sqft_basement.

```python
#Removing sqft_basement from the filtered dataframe
del subset_df['sqft_basement']
print(subset_df.columns)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

# Simple linear regression
Picking sqft_above as it has the highest correlation with sqft_living

```python
X_1 = subset_df[['sqft_above']]
y_1 = subset_df['sqft_living']
```

# Train & test data

```python
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_1, y_1, test_size=0.2, random_state=123)
```

# Model training

```python
simple_linear_model = LinearRegression().fit(X_train_1, y_train_1)
```

# Model testing

```python
x_test_value = X_test_1.iloc[1]
y_pred = simple_linear_model.predict([[1050]])
y_actual = y_test_1.iloc[1]
residual = y_pred - y_actual
print(residual)
```

# Evaluation

```python
print('Model coefficients', simple_linear_model.coef_)
print('Model intercept', simple_linear_model.intercept_)
r2_score_1 = simple_linear_model.score(X_test_1, y_test_1)
print('R square', r2_score_1)
print('Adjust R square', 1-(1-r2_score_1)*(len(X_train_1)-1)/(len(X_train_1)-1-1))
```

# Multi linear regression (Filtering out irrelevant fields)
1. id - Does not represent anything related to sqft_living
2. date, yr_built, yr_renovated - Time point would also play no role in the sqft_living of the house
3. zipcode, lat, long - Positional references also are not related to the sqft_living

```python
X_3 = df[[x for x in df.columns if x not in ['id', 'date', 'sqft_living', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long']]]
y_3 = df['sqft_living']
```

# Train & test

```python
X_train_3, X_test_3, y_train_3, y_test_3 = train_test_split(X_3, y_3, test_size=0.2, random_state=123)
```

# Model training
