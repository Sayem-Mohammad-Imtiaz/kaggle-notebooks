```python
class Crime():
    primary_type = ""
    description = ""
    freq = 0
    
    def __init__(self, primary_type, description, freq):
        self.primary_type = primary_type
        self.description = description
        self.freq = freq
    
    def __str__(self):
        crime_string = self.primary_type + " | " + self.description + " | " + str(self.freq)
        return crime_string
    
    def __cmp__(self, other):
        #if hasattr(other, 'number'):
        return self.freq.__cmp__(other.freq)

a = Crime("ha", "do", 3)
print(a)
```

```python
crime_list = []

i = 0
for key, col in cross.iterrows():
    #print(type(col))
    for d, c in col.items():
        crime_list.append(Crime(col.name, d, c))
        #print(crime_list[i])
        i += 1
```

```python
crime_list.sort(key=lambda x: x.freq, reverse=True)
for i in range(15):
    print(crime_list[i])
```
