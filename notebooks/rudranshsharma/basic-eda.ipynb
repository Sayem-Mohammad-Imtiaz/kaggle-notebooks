```python
import matplotlib.pyplot as plt
```

```python
X = pd.read_csv("/kaggle/input/hr-analytics-job-change-of-data-scientists/aug_test.csv")
X.head()
```

```python
a = X.isnull().sum() * 100/len(X)
missing = pd.DataFrame({'column_name': X.columns, 'percentage_missing':a })
missing
```

```python
import seaborn as sns
sns.heatmap(X.isnull())
```

```python
_, ax = plt.subplots(1, 3, figsize = (25, 5))
sns.countplot(X['relevent_experience'], hue = X['education_level'], ax = ax[0])
sns.countplot(X['experience'], ax = ax[1])
sns.countplot(X['experience']== '>20', hue = X['company_type'], ax = ax[2])
ax[0].set_xlabel('relevant experince splitted by education level');
ax[1].set_xlabel('relevant experince');
ax[2].set_xlabel('relevant experince > 20 splitted by the company type'); # only Intrsted in the >20 segment because of the sheer number of people here 


```

```python
_, ax = plt.subplots(1, 4, figsize = (25, 5))
sns.countplot(X['enrolled_university'], hue = X['education_level'], ax = ax[0])
sns.countplot(X['gender'], hue = X['education_level'] ,ax = ax[1]) # Male Dominace in the tech industry right here
sns.countplot(X['gender'], ax = ax[2])
sns.countplot(X['city_development_index'] > 0.5, hue = X['education_level'], ax = ax[3]) #shows most cities are indeed developed if the threshold would be 0.5
```

```python
_, ax = plt.subplots(1, 3, figsize = (20, 3))
sns.countplot(X[(X['education_level'] == 'Primary School')& (X['enrolled_university'] == "no_enrollment")]['experience'], ax = ax[0])
ax[0].set_xlabel('Distribution of people who did not have any education')
sns.barplot(y = [len(X[X['training_hours']>24]['training_hours']), len(X[(X['training_hours']<24)]['training_hours'])], x = ['>24', '<24'], ax = ax[1])
ax[1].set_xlabel('hours spent on training by the company')
sns.countplot(X['major_discipline'], ax = ax[2])
ax[2].set_xlabel('distribution of people based on the place where they had previously worked');
```
