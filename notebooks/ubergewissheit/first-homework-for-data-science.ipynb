```python
# Let's do plots! First, line pilot

data.Generosity.plot(color = 'g',label = 'Generosity',linewidth=1, grid = True,linestyle = '-')
plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('Countries')              # label = name of label
plt.ylabel('Generosity')
plt.title('Generosity across Countries')            # title = title of plot
plt.show()
```
