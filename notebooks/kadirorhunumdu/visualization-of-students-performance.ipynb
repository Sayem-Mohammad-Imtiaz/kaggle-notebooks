```python
#Average Scores by Race/Ethnicity 
race_names=data["race/ethnicity"].unique()
average_scores1=[]
average_scores2=[]
average_scores3=[]
for i in race_names:
    x=data[data["race/ethnicity"]==i]
    average_score=sum(x["math score"])/len(x)
    average_scores1.append(average_score)
    
for i in race_names:
    x=data[data["race/ethnicity"]==i]
    average_score=sum(x["reading score"])/len(x)
    average_scores2.append(average_score)

for i in race_names:
    x=data[data["race/ethnicity"]==i]
    average_score=sum(x["writing score"])/len(x)
    average_scores3.append(average_score)
    
new_data1=pd.DataFrame({"race_names":race_names,"average_scores1":average_scores1})
new_data2=pd.DataFrame({"race_names":race_names,"average_scores2":average_scores2})
new_data3=pd.DataFrame({"race_names":race_names,"average_scores3":average_scores3})

#Ascending
new_index1=new_data1["average_scores1"].sort_values(ascending=True).index
sorted_new1=new_data1.reindex(new_index1)
new_index2=new_data2["average_scores2"].sort_values(ascending=True).index
sorted_new2=new_data2.reindex(new_index2)
new_index3=new_data3["average_scores3"].sort_values(ascending=True).index
sorted_new3=new_data3.reindex(new_index3)


concat2=pd.concat([sorted_new1,sorted_new2["average_scores2"],sorted_new3["average_scores3"]],axis=1)
```

```python
concat2
```

```python
#Visualizations
trace1=go.Bar(x=concat2["race_names"],
              y=concat2["average_scores1"],
              name="Math Average Score",
              marker=dict(color="rgba(255,0,0,0.8)",
                         line=dict(color="rgb(0,0,0)",width=1.5)))

trace2=go.Bar(x=concat2["race_names"],
              y=concat2["average_scores2"],
              name="Reading Average Score",
              marker=dict(color="rgba(0,255,0,0.8)",
                         line=dict(color="rgb(0,0,0)",width=1.5)))

trace3=go.Bar(x=concat2["race_names"],
              y=concat2["average_scores3"],
              name="Writing Average Score",
              marker=dict(color="rgba(0,0,255,0.8)",
                         line=dict(color="rgb(0,0,0)",width=1.5)))

data8=[trace1,trace2,trace3]
layout=dict(title="Average Scores by Race/Ethnicity ",
           xaxis=dict(title="Race Names",ticklen= 5,zeroline= False),
           yaxis=dict(title="Average Scores",ticklen= 5,zeroline= False))
fig=dict(data=data8,layout=layout)
iplot(fig)
```

```python
#Average Scores by Test Preparation Course
test_pre=data["test preparation course"].unique()
average_scores11=[]
average_scores12=[]
average_scores13=[]

for i in test_pre:
    x=data[data["test preparation course"]==i]
    math_average=sum(x["math score"])/len(x)
    average_scores11.append(math_average)
    
for i in test_pre:
    x=data[data["test preparation course"]==i]
    reading_average=sum(x["reading score"])/len(x)
    average_scores12.append(reading_average)
    
for i in test_pre:
    x=data[data["test preparation course"]==i]
    writing_average=sum(x["writing score"])/len(x)
    average_scores13.append(writing_average)

n_data1=pd.DataFrame({"test_pre":test_pre,"average_scores11":average_scores11})
n_data2=pd.DataFrame({"test_pre":test_pre,"average_scores12":average_scores12})    
n_data3=pd.DataFrame({"test_pre":test_pre,"average_scores13":average_scores13}) 

concat3=pd.concat([n_data1,n_data2["average_scores12"],n_data3["average_scores13"]],axis=1)
```
