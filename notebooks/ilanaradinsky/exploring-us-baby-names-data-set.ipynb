```python
sammy = pd.DataFrame(national_names.loc[national_names["Name"] == "Sammy"])
sammy.head()
```

```python
sns.FacetGrid(sammy, hue="Gender", size=5) \
   .map(plt.scatter, "Year", "Count") \
   .add_legend()
```
