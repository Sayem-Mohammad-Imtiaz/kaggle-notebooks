```python
#splitting the data sets into independent (X) and dependent (Y) data sets
X = df.iloc[:,2:31].values
Y = df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
#scale the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(X_train)
sc.fit_transform(X_test)
```

```python
def models(X_train , Y_train):
    #logistic Regression
    from sklearn.linear_model import LogisticRegression
    log = LogisticRegression(random_state=0)
    log.fit(X_train,Y_train)
    
    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(criterion='entropy', random_state=0)
    tree.fit(X_train,Y_train)
    
    #Random Forest
    from sklearn.ensemble import RandomForestClassifier
    forest = RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
    forest.fit(X_train,Y_train)
    #KNN
    from sklearn.neighbors import KNeighborsClassifier
    KNN = KNeighborsClassifier(n_neighbors=5, weights='distance')
    KNN.fit(X_train,Y_train)
    
    #accuracy
    print("[0]Logistic Regression Training Accuracy: ",log.score(X_train,Y_train))
    print("[1]Decision Tree Training Accuracy: ",tree.score(X_train,Y_train))
    print("[2]Random Forest Training Accuracy: ",forest.score(X_train,Y_train))
    print("[3]KNN Classifier Training Accuracy: ",KNN.score(X_train,Y_train))
    
    return log,tree,forest,KNN
    
```

```python
model = models(X_train ,Y_train)
```

```python
#test model accuracy on test data on confusion matrix
from sklearn.metrics import confusion_matrix
for i in range(len(model)):
    print('Model ',i)
    cm = confusion_matrix(Y_test,model[i].predict(X_test))
    TP = cm[0][0]
    TN = cm[1][1]
    FP = cm[1][0]
    FN = cm[0][1]
    
    print(cm)
    print('Testing Accuracy =', (TP+TN)/(TP + TN + FP + FN))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
for i in range(len(model)):
    print('Model ',i)
    print(classification_report(Y_test,model[i].predict(X_test)))
    print("Accuracy =",accuracy_score(Y_test,model[i].predict(X_test)))
    print()
```
