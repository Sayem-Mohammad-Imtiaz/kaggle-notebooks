```python
covid = pd.read_csv("../input/corona-virus-report/country_wise_latest.csv")
covid2 = covid[['Country/Region','Confirmed','Deaths','Recovered','Active']].copy()
covid.head()
```

```python
covid_asc = covid2.sort_values(by='Confirmed', ascending=False)
covid_asc.iloc[:10].style.background_gradient(cmap='coolwarm')
```

```python
sns.set(style='whitegrid')

from pylab import rcParams
rcParams['figure.figsize'] = 10, 7

covid_asc.iloc[0:10].plot(x='Country/Region',y='Confirmed',kind='bar',color='Blue')
plt.legend(['Toal de casos confirmados'], loc='upper right')
plt.title("Total de Casos confirmados por Pais",fontdict = {"fontsize":20})
plt.xlabel("Países")
plt.ylabel("Casos confirmados")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
Active = np.array([[281644],[508116],[254427],[47657],[12581],[495499],[108928],[93613],[98752],[25550]]).reshape((-1, 1))
print(Active)
```

```python
Deaths = np.array([[148011],[87618],[45844],[44022],[35112],[33408],[30212],[28432],[18418],[15912]]).reshape((-1, 1))
print(Deaths)
```

```python
%matplotlib inline

plt.scatter(Active, Deaths)
plt.title('Virus ativo X Mortos')
plt.xlabel('Mortos')
plt.ylabel('Virus ativo')
```

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(Active, Deaths)
```

```python
previsoes = regression.predict(Deaths)

from sklearn.metrics import mean_absolute_error, mean_squared_error
morte = mean_absolute_error(Active, previsoes)
print('morte: ', morte)
média = mean_squared_error(Active, previsoes)
print('média: ', média)
```

```python
plt.plot(Deaths, Active, 'o')
plt.plot(Deaths, previsoes, color='Red')
plt.title('Previsão de Mortes')
plt.xlabel('Mortos')
plt.ylabel('Vírus Ativo')
```
