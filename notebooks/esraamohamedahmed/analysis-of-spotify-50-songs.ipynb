```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from scipy import stats
import squarify as sq
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import seaborn as sns
import sklearn
%matplotlib inline
```

```python
file='/kaggle/input/top50spotify2019/top50.csv'
spotify=pd.read_csv(file,encoding='ISO-8859-1')
spotify.head()
```

```python
fig, axs = plt.subplots(5,2,figsize=(15,15))
axs[0, 0].hist(spotify['Beats.Per.Minute'],bins=10) #original data
axs[0, 0].set_title('Beats.Per.Minute')
axs[0, 1].hist(spotify['Energy'],bins=10) 
axs[0, 1].set_title('Energy')
axs[1, 0].hist(spotify['Danceability'],bins=10) 
axs[1, 0].set_title('Danceability')
axs[1, 1].hist(spotify['Loudness..dB..'],bins=10) 
axs[1, 1].set_title('Loudness..dB..')

axs[2, 0].hist(spotify['Liveness'],bins=10) 
axs[2, 0].set_title('Liveness')

axs[2, 1].hist(spotify['Valence.'],bins=10) 
axs[2, 1].set_title('Valence.')

axs[3, 0].hist(spotify['Length.'],bins=10)
axs[3, 0].set_title('Length.')

axs[3, 1].hist(spotify['Acousticness..'],bins=10)
axs[3, 1].set_title('Acousticness..')

axs[4, 0].hist(spotify['Speechiness.'],bins=10) 
axs[4, 0].set_title('Speechiness.')

axs[4, 1].hist(spotify['Popularity'],bins=10) 
axs[4, 1].set_title('Popularity')

#plt.show()
#axs[0, 0].set_title('Axis [0, 0]')
#axs[0, 1].hist(spotify['Energy'],bins=10) #original data
plt.show()



```

```python
plt.figure(figsize=(12,12))
sns.jointplot(x=spotify["Danceability"].values, y=spotify['Popularity'].values, size=12, kind="reg",truncate=False,)
plt.ylabel('Popularity', fontsize=12)
plt.xlabel("Danceability", fontsize=12)
plt.title("Danceability Vs Popularity", fontsize=15)
plt.show();
```

```python
plt.figure(figsize=(12,12))
sns.jointplot(x=spotify["Speechiness."].values, y=spotify['Popularity'].values, size=12, kind="reg",truncate=False,)
plt.ylabel('Popularity', fontsize=12)
plt.xlabel("Speechiness.", fontsize=12)
plt.title("Speechiness Vs Popularity", fontsize=15)
plt.show();
```

```python
plt.figure(figsize=(12,12))
sns.jointplot(x=spotify["Speechiness."].values, y=spotify['Beats.Per.Minute'].values, size=12, kind="reg",truncate=False,)
plt.ylabel('Beats.Per.Minute', fontsize=12)
plt.xlabel("Speechiness.", fontsize=12)
plt.title("Speechiness Vs Beats.Per.Minute", fontsize=15)
plt.show();
```
