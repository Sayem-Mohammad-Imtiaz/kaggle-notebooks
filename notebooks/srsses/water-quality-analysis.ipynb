# So we have 3 columns containing null values and we can either remove the rows or fill them with zeros or any appropriate value
# But since it is a qulity check so removing all or replacing with zeros can cause biased result so replacing them with te mean values will be helpfull for the analysis

```python
water_new['ph'] = water_new['ph'].fillna(water_new['ph'].mean())
water_new['Sulfate']=water_new['Sulfate'].fillna(water_new['Sulfate'].mean())
water_new['Trihalomethanes']=water_new['Trihalomethanes'].fillna(water_new['Trihalomethanes'].mean())
water_new.isnull().sum()
```

So we can see now we have filled all the Nan Values according to the mean of respective column

# So as we know the data frame has values which might be dangerous if present in excess or can make water un usable. So we will check the PH value first to see how the water quality is regarding PH values. The maximum PH of drinking water should be 8.5 so we will only keep the water whose PH level is less than or equal to 8.5

```python
Ph = water_new[water_new['ph']<=8.5]
Ph
```

So we have 489 rows having PH value more than the recomended level
