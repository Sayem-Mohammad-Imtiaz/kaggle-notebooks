```python
not_A = df[df['Attrition_yes']==0]
yes_A = df[df['Attrition_yes']==1]
```

```python
from sklearn.utils import resample
upsampx = resample(yes_A,replace=True,n_samples=len(not_A))
```

```python
upsamp = pd.concat([not_A,upsampx])
```

```python
upsamp['Attrition_yes'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
for i in df.columns:
    if df[i].dtype == np.int64:
        continue
    else:
        df[i] = LabelEncoder().fit_transform(df[i])
```

```python
#Data PreProcessing
```

```python
x= df.iloc[:,:-1]
y = df['Attrition_yes']
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
```

```python
xcol = x.columns

x = pd.DataFrame(sc.fit_transform(x),columns=xcol)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(x_train,y_train)
```

```python
y_lr = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_lr,y_test)
```
