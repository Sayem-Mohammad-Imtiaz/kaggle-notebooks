```python
# Removing the extra column at the end and the id column (id doesn't effect the diagnosis)
df = df.drop(['Unnamed: 32'], axis=1)
df = df.drop(['id'], axis=1)
df.head(3)
```


******** Correlation analysis (correlation matrix & correlation heatmap) can be added here ********


```python
# Diagnosis' data type is object(string). We convert it into 1/0
df.diagnosis = [1 if i == 'M' else 0 for i in df.diagnosis]
```

```python
# Defining X as parameters and y as labels
X = df.drop(['diagnosis'], axis=1)
y = df.diagnosis.values
```

```python
# Normalization
# We use min/max method
X = (X - np.min(X)) / (np.max(X) - np.min(X))
```

```python
# Splitting train,test and X,y
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
d_tree = DecisionTreeClassifier()
d_tree.fit(X_train, y_train)
d_tree.score(X_test, y_test)
```
