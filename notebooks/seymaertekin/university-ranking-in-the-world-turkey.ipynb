The following columns are the performance indicators for the total score and rank; 

**teaching** is the indicator of the learning environment
1. Reputation survey: 15%
2. Staff-to-student ratio: 4.5%
3. Doctorate-to-bachelor’s ratio: 2.25%
4. Doctorates awarded-to-academic staff ratio: 6%
5. Institutional income: 2.25%

**research** is the indicator of volume, income and reputation
1. Reputation survey: 18%
2. Research income: 6%
3. Research productivity: 6%

**citations** is the indicator of research influence

**international** is the indicator of staff, students and research
1. International-to-domestic-student ratio: 2.5%
2. International-to-domestic-staff ratio: 2.5%
3. International collaboration: 2.5%

**income** is the indicator of knowledge transfer
A university’s ability to help industry with innovations, inventions and consultancy
