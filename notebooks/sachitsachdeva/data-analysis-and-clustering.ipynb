# CLUSTERING

```python
X = dataset.drop(['State/UTs','Active Ratio (%)','Discharge Ratio (%)','Death Ratio (%)'],axis=1).iloc[:,:].values
```

```python
print(X)
```

ELBOW METHOD

```python
from sklearn.cluster import KMeans 
wcss = []
for i in range(1,10):
  kmeans = KMeans(n_clusters = i,init = 'k-means++',random_state = 42 )
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
plt.plot(range(1,10),wcss)
plt.title("ELBOW METHOD")
plt.xlabel('NUMBEROF CLUSTERS')
plt.ylabel('WCSS')
plt.show()
```

3 CLUSTERS

```python
kmeans = KMeans(n_clusters = 3,init = 'k-means++',random_state = 42)
y_kmeans = kmeans.fit_predict(X)
```

```python
print(y_kmeans)
```

# ANALYSING THE CLASSES FORMED

```python
newdataset = dataset
newdataset['y_kmeans']=y_kmeans
```

```python
newdataset[newdataset['y_kmeans']==0]['Total Cases'].describe()
```
