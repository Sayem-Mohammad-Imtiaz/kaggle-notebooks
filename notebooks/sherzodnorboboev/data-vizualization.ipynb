# 1. Bar Plot

<a id="2"></a><br>
## Poverty rate of each state

```python
area_list = list(below_p['Geographic Area'].unique())
area_powerty_ratio = []

for i in area_list:
    x = below_p[below_p['Geographic Area'] == i]
    area_powerty_rate = mean(x.poverty_rate)
    area_powerty_ratio.append(area_powerty_rate)

data = pd.DataFrame({'area_list': area_list, 'area_powerty_ratio': area_powerty_ratio})
new_index = (data.area_powerty_ratio.sort_values(ascending=False)).index.values
sorted_data = data.reindex(new_index)

# vis
plt.figure(figsize=(15,10))
sns.barplot(x=sorted_data['area_list'],y=sorted_data['area_powerty_ratio'])
plt.xticks(rotation = 60)
plt.xlabel('States')
plt.ylabel('Powerty ratio')
plt.title('Powerty given states')
plt.show()
```

## Most common 15 Name or Surname of killed people

```python
# Most common 15 name or surname of killed people
separate = killed.name[killed.name != 'TK TK'].str.split()
a, b = zip(*separate)
name_list = a + b
name_count = Counter(name_list)
most_common_names = name_count.most_common(15)
x, y = zip(*most_common_names)
x, y = list(x), list(y)

# hue=None, data=None, order=None, hue_order=None, 
# estimator=<function mean at 0x000001AF9BDD6670>, 
# ci=95, n_boot=1000, units=None, seed=None, 
# orient=None, color=None, palette=None, saturation=0.75, 
# errcolor='.26', errwidth=None, capsize=None, dodge=True

plt.figure(figsize=(15, 10))
sns.barplot(x=x, y=y, palette = sns.cubehelix_palette(len(x), reverse=True), saturation=0.9)
plt.xlabel('Name or Surname of killed people')
plt.ylabel('Frequency')
plt.title('Most common 15 name or Surname of killed people')
plt.show()
```

## High school graduation rate of the population that is older than 25 in states

```python
high_s.percent_completed_hs = high_s.percent_completed_hs.astype('float')
aria_list = high_s['Geographic Area'].unique()
area_highschool = []
for i in aria_list:
    x = high_s[high_s['Geographic Area'] == i]
    area_highschool_rate = mean(x.percent_completed_hs)
    area_highschool.append(area_highschool_rate)

# sorting
data = pd.DataFrame({ 'aria_list': aria_list, 'area_highschool_ratio': area_highschool })
new_index = (data.area_highschool_ratio.sort_values(ascending=True)).index.values
sorted_data2 = data.reindex(new_index)

# viz
plt.figure(figsize=(15,10))
sns.barplot(x=sorted_data2.aria_list, y=sorted_data2.area_highschool_ratio)
plt.xticks(rotation=60)
plt.xlabel('States')
plt.ylabel('High_School Graduate Rate')
plt.title("Percentage of Given State's Population Above 25 that Has Graduated High School")
plt.show()
```

## Percentage of state's population according to races that are black,white,native american, asian and hispanic

```python
# share_r.replace(['-'], 0.0, inplace=True)
share_r.replace(["(X)"], 0.0, inplace=True)
share_r.share_white = share_r.share_white.astype('float')
share_r.iloc[:, 2:] = share_r.iloc[:, 2:].astype(float)
aria_list = share_r['Geographic area'].unique()

share_white           = []
share_black           = []
share_native_american = []
share_asian           = []
share_hispanic        = []

for i in aria_list:
    x = share_r[share_r['Geographic area'] == i]
    share_white.append(mean(x.share_white))
    share_black.append(mean(x.share_black))
    share_native_american.append(mean(x.share_native_american))
    share_asian.append(mean(x.share_asian))
    share_hispanic.append(mean(x.share_hispanic))

# vizualization 
f, ax = plt.subplots(figsize = (10, 15))
# (<Figure size 1080x1080 with 1 Axes>,                       = f
#  <matplotlib.axes._subplots.AxesSubplot at 0x7f56a1c0aa90>) = ax

sns.barplot(x = share_white,           y = aria_list, color = 'green',  alpha = 0.8, label = 'White')
sns.barplot(x =share_black,            y = aria_list, color = 'blue',   alpha = 0.7, label = 'Black')
sns.barplot(x =share_native_american,  y = aria_list, color = 'yellow', alpha = 0.8, label = 'Native American')
sns.barplot(x =share_asian,            y = aria_list, color = 'cyan',   alpha = 0.8, label = 'Asian')
sns.barplot(x =share_hispanic,         y = aria_list, color = 'red',    alpha = 0.6, label = 'Hispanic')

ax.legend(loc='lower right', frameon=True) # legendary's alpha
ax.set(xlabel='Percentage of Races', ylabel = 'States', title = "Percentage of State's population")
plt.show()
```

## High school graduation rate vs Poverty rate of each state

```python
# high school graduation rate vs Poverty rate of each state
sorted_data['area_powerty_ratio'] = sorted_data['area_powerty_ratio'] / max(sorted_data['area_powerty_ratio'])
sorted_data2['area_highschool_ratio'] = sorted_data2['area_highschool_ratio'] / max(sorted_data2['area_highschool_ratio'])

data = pd.concat([sorted_data, sorted_data2['area_highschool_ratio']], axis = 1)
data.sort_values('area_powerty_ratio', inplace = True)

# Vizualization
f, ax = plt.subplots(figsize = (20, 10))
sns.pointplot(x = 'area_list', y = 'area_powerty_ratio',    data = data, color = 'lime', label = 'Area Powert Ratio')
sns.pointplot(x = 'area_list', y = 'area_highschool_ratio', data = data, color = 'red',  label = 'Area Highschool Ratio')
plt.text(40, 0.35, 'high school graduation ratio', color = 'red', fontsize = 16, style = 'italic')
plt.text(40, 0.3,  'poverty ratio', color = 'lime', fontsize = 16, style = 'italic')
plt.xlabel('States', fontsize = 15, color = 'blue')
plt.ylabel('Values', fontsize = 15, color = 'blue')
plt.title('High School Graduate VS Povert Ratio', fontsize = 20, color = 'blue')
plt.grid()
```
