```python
##traitement des données
df['log_price']=np.log(1+df['price'])
df['log_sqft_lot']=np.log(1+df['sqft_lot'])
df = pd.get_dummies(df,columns=['view'])
df = pd.get_dummies(df,columns=['condition'])
```

```python
df.columns
```

```python
df=df[['log_price','log_sqft_lot','sqft_living','floors','waterfront','grade','bedrooms','bathrooms','view_0','view_1','view_2','view_3','view_4','condition_1','condition_2', 'condition_3', 'condition_4', 'condition_5']]
```

```python
df.columns
```

```python
train_df=df[:17000] #80% des données apprentissage
test_df=df[17000:] #20% des données pour le test
train_X=train_df[['log_sqft_lot', 'sqft_living', 'floors', 'waterfront','grade', 'bedrooms', 'bathrooms', 'view_0', 'view_1', 'view_2','view_3', 'view_4', 'condition_1', 'condition_2', 'condition_3','condition_4', 'condition_5']].values
train_Y=train_df[['log_price']].values
test_X=test_df[['log_sqft_lot', 'sqft_living', 'floors', 'waterfront','grade', 'bedrooms', 'bathrooms', 'view_0', 'view_1', 'view_2','view_3', 'view_4', 'condition_1', 'condition_2', 'condition_3','condition_4', 'condition_5']].values
test_Y=test_df[['log_price']].values
print( train_X.shape, test_X.shape )
print( train_Y.shape, test_Y.shape )
```

```python
##application de min max pour les données d'apprentissage et de test
scaler = MinMaxScaler()
train_X = scaler.fit_transform(train_X)
test_X = scaler.transform(test_X)
```

```python
inpt   = layers.Input( shape=(17,))
x      = layers.Dense(300) ( inpt )
actv   = layers.Activation('relu') (x)
output = layers.Dense(1) (actv)
model  = Model( inpt, output )
model.summary()
```

```python
opt = SGD()
model.compile(optimizer=opt,loss='mse') 
model.fit(train_X,train_Y,epochs=10)
```

```python
prediction = model.predict(test_X)
prediction
np.exp(prediction)-1
```
