```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model
```

```python
df = pd.read_csv('/kaggle/input/predict-test-scores-of-students/test_scores.csv').set_index('student_id')
print(df.describe())
df.head()
```

```python
df.dtypes
```

```python
df['school'] = df['school'].astype('category')
df['school_setting'] = df['school_setting'].astype('category')
df['school_type'] = df['school_type'].astype('category')
df['classroom'] = df['classroom'].astype('category')
df['teaching_method'] = df['teaching_method'].astype('category')
df['gender'] = df['gender'].astype('category')
df['lunch'] = df['lunch'].astype('category')
df.dtypes
```

```python
cat_columns = df.select_dtypes(['category']).columns
df[cat_columns] = df[cat_columns].apply(lambda x: x.cat.codes)
```

```python
df.head()
```

```python
X = df.drop('posttest', axis=1)
y = df['posttest']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
reg = linear_model.LinearRegression()
reg.fit(X_train, y_train)
```

```python
reg.score(X_test, y_test) #RÂ² score
```

```python
print('Some predictions :')
for i in range(10):
    predict = reg.predict(np.array(X_test.iloc[i]).reshape(-1,9))
    print(f'Target={y_test.iloc[i]}  Prediction={predict[0]}')
```
