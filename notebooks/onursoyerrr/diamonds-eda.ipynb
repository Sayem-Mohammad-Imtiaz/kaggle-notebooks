**Content**

**price** price in US dollars (\$326--\$18,823)

**carat** weight of the diamond (0.2--5.01)

**cut** quality of the cut (Fair, Good, Very Good, Premium, Ideal)

**color** diamond colour, from J (worst) to D (best)

**clarity** a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))

**x** length in mm (0--10.74)

**y **width in mm (0--58.9)

**z** depth in mm (0--31.8)

**depth** total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)

**table** width of top of diamond relative to widest point (43--95)

```python
#reading data
data = pd.read_csv("/kaggle/input/diamonds/diamonds.csv")
data.head()
```

```python
data.info()
```

```python
data.corr()
```

This code section shows that the relations of columns in data. If value is 1 or almost 1 we can say there is a quite relation between these two columns. 

```python
#correlation map 
f,ax = plt.subplots(figsize = (13,13))
sns.heatmap(data.corr(),annot = True , linewidth = 5, fmt = '.1f' , ax = ax)
plt.show()
#visualising the correlations between columns 
```

```python
#line plot 
data.price.plot(kind = 'line' , label = 'Price' , grid = True, alpha = 0.5 , linewidth = 1 , linestyle = '-.',color = 'r')
data.carat.plot(label = 'Carat', grid = True , alpha = 0.5 , linewidth = 1 , linestyle = '-.', color = 'g')
plt.legend(loc = 'upper right')
plt.xlabel('Price')                    
plt.ylabel('Carat')
plt.title('Line Plot')
plt.show()
```
