Ambil data dengan syarat:
1. Umur >= 25
2. Tinggi badan >=200
3. Berat badan >= 90
4. Point >= 10
5. Gp >=50

```python
#ambil colom sesuai syarat
df = NBA[["age","player_height","player_weight","pts","gp"]]
```

```python
df.head()
```

```python
#membuat dataset sesuai dengan syarat diatas
def dataset(x):
  if x['age']>=25 and x['player_height']>=200 and x['player_weight']>=90 and x['pts']>=10 and x['gp']>=50:
    return 1
  else:
    return 0

df['keterangan']=df.apply(dataset,axis=1)
```

```python
df.head(30)
```

```python
df[df['keterangan']==1].head(100)
```

Standardscaler digunakan untuk menghilangkan mean dan menskalakan ke variasi. biasanya standardscaler digunakan untuk membagi data menjadi train dan test.

rujukan:
1. https://prasetyadi.name/2019/data-preprocessing-menggunakan-modul-python/
2. https://qastack.id/datascience/38395standardscaler-before-and-after-splitting-data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
val= sc.fit_transform(df)
```

```python
#membagi dataset ke dalam training set dan test set (untuk kasus ini diambil 30% untuk test)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df, df['keterangan'], test_size= .30)
len(x_train)
```

```python
#load machine learning
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import Perceptron
from sklearn.linear_model import SGDClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC,LinearSVC
from sklearn.naive_bayes import GaussianNB
```

```python
sgd = linear_model.SGDClassifier(max_iter=5, tol=None)
sgd.fit(x_train, y_train)
y_pred = sgd.predict(x_test)

sgd.score(x_train, y_train)

acc_sgd = round(sgd.score(x_train, y_train) * 100, 2)
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(x_train,y_train)
y_pred=random_forest.predict(x_test)

random_forest.score(x_train,y_train)
acc_random_forest=round(random_forest.score(x_train,y_train)*100,2)
```

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)

y_pred = logreg.predict(x_test)

acc_log = round(logreg.score(x_train, y_train) * 100, 2)
```

```python
KNN = KNeighborsClassifier(n_neighbors=3)
KNN.fit(x_train,y_train)
y_pred = KNN.predict(x_test)
acc_KNN = round(KNN.score(x_train,y_train)*100,2)
```

```python
perceptron = Perceptron()
perceptron.fit(x_train, y_train)

y_pred = perceptron.predict(x_test)

acc_perceptron = round(perceptron.score(x_train, y_train) * 100, 2)
```

```python
linear_svc =LinearSVC()
linear_svc.fit(x_train,y_train)

y_pred = linear_svc.predict(x_test)

acc_linear_svc = round(linear_svc.score(x_train,y_train)*100,2)
```

```python
gaussian = GaussianNB() 
gaussian.fit(x_train, y_train)  
y_pred = gaussian.predict(x_test)  
acc_gaussian = round(gaussian.score(x_train, y_train) * 100, 2)
```

```python
decision_tree = DecisionTreeClassifier() 
decision_tree.fit(x_train, y_train)  
Y_pred = decision_tree.predict(x_test)  
acc_decision_tree = round(decision_tree.score(x_train, y_train) * 100, 2)
```

```python
results = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 
              'Random Forest', 'Naive Bayes', 'Perceptron', 
              'Stochastic Gradient Decent', 
              'Decision Tree'],
    'Score': [acc_linear_svc, acc_KNN, acc_log, 
              acc_random_forest, acc_gaussian, acc_perceptron, 
              acc_sgd, acc_decision_tree]})
result_df = results.sort_values(by='Score', ascending=False)
result_df = result_df.set_index('Score')
result_df.head(9)
```

```python
random_forest = RandomForestClassifier(criterion = "gini", 
                                       min_samples_leaf = 1, 
                                       min_samples_split = 10,   
                                       n_estimators=100, 
                                       max_features='auto', 
                                       oob_score=True, 
                                       random_state=1, 
                                       n_jobs=-1)

random_forest.fit(x_train, y_train)
Y_prediction = random_forest.predict(x_test)

random_forest.score(x_train, y_train)

print("oob score:", round(random_forest.oob_score_, 4)*100, "%")
```
