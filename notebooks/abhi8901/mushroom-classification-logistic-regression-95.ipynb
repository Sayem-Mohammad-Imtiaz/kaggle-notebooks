```python
from sklearn.preprocessing import LabelEncoder
def label_encoded(x):
    le = LabelEncoder()
    le.fit(x)
    print(x.name, le.classes_)
    return le.transform(x)
```

```python
for col in df.columns:
    df[str(col)] = label_encoded(df[str(col)])
```

# Training data using Random Forest Classifier

```python
train = df.drop(['class'], axis=1)
test = df['class']
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(train, test, test_size=0.33, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

model_1 = RandomForestClassifier(max_depth=10, random_state=0)
```

```python
model_1.fit(train_x, train_y)
```

```python
yhat = model_1.predict(test_x)
```

```python
accuracy_score(test_y, yhat)
```

# Training data using Logistic Regression

```python
data = df.values
x = data[:, 1:-1].astype(str)
y = data[:, 0].astype(str)
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.33, random_state=0)
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
ordinal_encoder.fit(train_x)
train_x = ordinal_encoder.transform(train_x)
test_x = ordinal_encoder.transform(test_x)
# ordinal encode target variable
label_encoder = LabelEncoder()
label_encoder.fit(train_y)
train_y = label_encoder.transform(train_y)
test_y = label_encoder.transform(test_y)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(solver='liblinear')
LR.fit(train_x, train_y)
y_pred = LR.predict(test_x)
accuracy_score(test_y, y_pred)
```
