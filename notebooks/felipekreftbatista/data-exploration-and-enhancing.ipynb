```python
space['company_nature'] = 'private'
space['company_hq'] = '---'
print(space.company_name.value_counts())
```

with so many companies, it would be good to create a function that assigns the nature and hq to a given company

And also 2 more counters, one with the unmarked companies and the other with the marked companies

```python
spacecopy = space.copy()

def company_nature(company_name, nature, hq):
    ''' company name, nature, HQ location'''
    index = space.query('company_name == "{}"'.format(company_name)).index
    space['company_nature'][index] = '{}'.format(nature)
    space['company_hq'][index] = '{}'.format(hq)
    print('function result \n',space.query('company_name == "{}"'.format(company_name)).company_nature.head(2),'\n')
    
    companies.pop(0)
    companies_marked.append('{}'.format(company_name))
    print('next company --- ', companies[0], ' ---')
    return 

companies = list(space.company_name.unique())
companies_marked = list()

print('companies: ', companies, '\n\n size of companies: ', len(companies))
print('companies marked: ', companies_marked, '\n\n size of companies marked: ', len(companies_marked))
```
