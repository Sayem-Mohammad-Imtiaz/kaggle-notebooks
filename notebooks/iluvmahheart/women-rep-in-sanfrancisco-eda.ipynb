## Visualization

```python
m = 0
f = 0
f_m = 0
m_f = 0
m_m = 0

for i,row in df.iterrows():
    if row['Gender'] == 'M':
        m+=1 
    
    elif row['Gender'] == 'F':
        f+=1
    
    elif row['Gender'] == 'F & M':
        f_m+=1
    
    elif row['Gender'] == 'M & F':
        m_f+=1
    
    elif row['Gender'] == 'M & M':
        m_m+=1

        
print("m:" , m , "f:" ,f ,  "f_m:" ,f_m  , "m_f:" , m_f ,  "m_m:" , m_m  )
```

```python
sns.countplot(x = "Gender" , data = df)
```

```python
sns.countplot(y = "Department/Source" ,hue ="Gender", data = df)
```
