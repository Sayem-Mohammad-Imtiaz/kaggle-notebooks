```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Import Data
Importing data and viewing data file information.

```python
dataset = pd.read_csv("../input/daily-power-generation-in-india-20172020/file_02.csv")
dataset=dataset.drop(['index'], axis=1)
dataset.info()
dataset
```

# Data Types
Column names have been changed. Incorrect data types corrected.
* date:          object -> datetime64[ns]
* therm_gen_act: objecy -> float64
* therm_gen_est: objecy -> float64

```python
dataset.columns=["date","region","therm_gen_act","therm_gen_est","nu_gen_act","nu_gen_est","hyd_gen_act","hyd_gen_est"]
dataset['date']= pd.to_datetime(dataset['date'])
dataset["therm_gen_act"]=dataset["therm_gen_act"].apply(lambda x:str(x).replace(',','')if ',' in str(x) else str(x))
dataset["therm_gen_act"]=dataset["therm_gen_act"].apply(lambda x: float(x))
dataset["therm_gen_est"]=dataset["therm_gen_est"].apply(lambda x:str(x).replace(",","")if ',' in str(x) else str(x))
dataset["therm_gen_est"]=dataset["therm_gen_est"].apply(lambda x: float(x))
dataset.info()
```

# Empty Values
A value of "0" was entered in the blank cells in the Nuclear Generation column.

```python
dataset["nu_gen_act"]=dataset.nu_gen_act.fillna(0.0)
dataset["nu_gen_est"]=dataset.nu_gen_est.fillna(0.0)
dataset
```

# Dataset Description
Dataset description (count, mean, min, max)

```python
dataset.describe()
```

# Total Generation
Total Generations of 3 years.

```python
dataset.describe().sum()
```

```python
daily_gen = dataset.groupby("date").sum()
daily_gen
```

```python
plt.figure(figsize=(20,8))
plt.plot(daily_gen.therm_gen_act, label= 'Thermal Generation Actual (in MU)', linewidth=1)
plt.plot(daily_gen.therm_gen_est, label= 'Thermal Generation Estimated (in MU)',linewidth=1)
plt.title("Thermal Power Plant Total Daily Generation in the Country.")
plt.xlabel("Dates")
plt.ylabel("Generation")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.plot(daily_gen.nu_gen_act, label= 'Nuclear Generation Actual (in MU)', linewidth=1)
plt.plot(daily_gen.nu_gen_est, label= 'Nuclear Generation Estimated (in MU)',linewidth=1)
plt.title("Nuclear Power Plant Total Daily Generation in the Country.")
plt.xlabel("Dates")
plt.ylabel("Generation")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20,8))
plt.plot(daily_gen.hyd_gen_act, label= 'Hydro Generation Actual (in MU)', linewidth=1)
plt.plot(daily_gen.hyd_gen_est, label= 'Hydro Generation Estimated (in MU)',linewidth=1)
plt.title("Hydro Power Plant Total Daily Generation in the Country.")
plt.xlabel("Dates")
plt.ylabel("Generation")
plt.legend()
plt.show()
```
