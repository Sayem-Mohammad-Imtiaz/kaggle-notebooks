# Catching the dataset

```python
# data
path = '/kaggle/input/heart-attack-analysis-prediction-dataset/heart.csv' # data
```

```python
# open data
heart_df = pd.read_csv(path) 
```

# Looking for missing values

#### Here we can see dataset does not have missing values.

```python
# missing values
heart_df.isnull().sum()
```

#### Replacing dummies values to categorical

```python
# setting values on sex column
sex_data = { 1: 'male', 0: 'female'}
heart_df['sex'] = heart_df['sex'].replace(sex_data)
sex = heart_df['sex']
count_sex= sex.value_counts()
```

# Gender pattients

#### The dataset is not distributed equally by gender. Most patients are male (68,3%). This could result in a biased analysis.

```python
# plotting the sex percentages
count_sex.plot(kind='pie', autopct="%.1f%%")
```

# Age's Pattients

#### The histogram shows the most frequent ages are between 57 and 61 years old.

```python
# setting age columns and plotting
age = heart_df['age']
count, bin_edges = np.histogram(heart_df['age'])
heart_df['age'].plot(kind='hist', xticks=bin_edges)
plt.xlabel('Age')
plt.title("Patient's Ages")
age.describe()
```

```python
# converting the column Chest Pain into object
for col in ['cp',]:
    heart_df[col] = heart_df[col].astype('object')
```

```python
# replacing the int values to categorical 
cp_data = { 0: 'typical angina', 1: 'atypical angina', 2: 'non-anginal pain', 3: 'asymptomatic'}
heart_df['cp'] = heart_df['cp'].replace(cp_data)
restecg_data = { 0: 'normal', 1: 'ST-T wave abnormality', 2: 'left ventricular hypertrophy'}
heart_df['restecg'] = heart_df['restecg'].replace(restecg_data)
```

```python
# converting columns into boolean value
for col in ['fbs', 'exng']:
    heart_df[col] = heart_df[col].astype('bool')
```

# Looking for correlation

#### The heatmap shows there is no strong correlation between the variables in the dataset.

```python
# setting values to correlation
feat = ['age', 'trtbps', 'chol', 'fbs', 'thalachh', 'exng']
corr = heart_df[feat].corr(method='pearson')
```

```python
# plotting heatmap to correlation
sns.heatmap(corr,xticklabels=corr.columns, yticklabels=corr.columns, cmap='coolwarm')
```
