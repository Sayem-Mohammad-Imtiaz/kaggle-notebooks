# Visualising happiness across the world in 2021

*Examining data about happiness, freedom, GDP, social support and life expectancy in different countries*

```python
import pandas as pd
from matplotlib import pyplot
```

After setting up and importing all necessary libraries, extract variables from each column of the data.

```python
# Extract column data
df = pd.read_csv('../input/world-happiness-report-2021/world-happiness-report-2021.csv')
happiness_score = df['Ladder score'].tolist()
gdp = df['Logged GDP per capita'].tolist()
social_support = df["Social support"].tolist()
life_expectancy = df["Healthy life expectancy"].tolist()
freedom = df["Freedom to make life choices"].tolist()
```

Create a scatter chart of freedom against happiness in high and low GDP countries.

```python
# Scatter chart of freedom and happiness in high and low GDP countries
fig = pyplot.figure(figsize=(8, 6))
scatter = pyplot.scatter(freedom, happiness_score, c=gdp, cmap="viridis")
legend = pyplot.legend(*scatter.legend_elements(num=5), title="Logged GDP per head")
pyplot.title("Freedom and happiness in high and low GDP countries")
pyplot.xlabel("Freedom")
pyplot.ylabel("Happiness Score")
pyplot.show()
```

Create a scatter chart of social support and life expectancy in countries with varying happiness levels.

```python
# Scatter chart of social_support and life_expectancy in happy/unhappy countries
fig = pyplot.figure(figsize=(8, 6))
scatter = pyplot.scatter(social_support, life_expectancy, c=happiness_score, cmap="plasma")
legend = pyplot.legend(*scatter.legend_elements(num=5), title="Happiness Score")
pyplot.title("Social support and life expectancy in happy/unhappy countries")
pyplot.xlabel("Social support")
pyplot.ylabel("Life expectancy")
pyplot.show()
```
