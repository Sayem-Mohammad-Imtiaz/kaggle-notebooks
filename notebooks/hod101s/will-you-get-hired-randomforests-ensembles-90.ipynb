```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, plot_confusion_matrix, confusion_matrix
```

```python
data = pd.read_csv('../input/factors-affecting-campus-placement/Placement_Data_Full_Class.csv')
data.head()
```

# Label Encoded DataFrame

```python
lbldata = data.copy()
lbldata.drop(['sl_no','salary'],axis=1,inplace=True)
categorical = ['gender','ssc_b','hsc_b','hsc_s','degree_t','workex','specialisation','status']
for col in categorical:
    lbldata[col]  = LabelEncoder().fit_transform(lbldata[col])
lbldata.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(lbldata.drop('status',axis=1),lbldata.status,test_size=0.2)
```

# Random Forests

```python
parameters = {
    'n_estimators'      : [320,330,340],
    'max_depth'         : [8,16,32,64],
    'random_state'      : [0,32,42]
}

clf = GridSearchCV(RandomForestClassifier(), parameters, cv=10, n_jobs=-1)
clf.fit(X_train,y_train)
```

```python
preds = clf.predict(X_test)
print(classification_report(y_test,preds))
```

```python
plot_confusion_matrix(clf,X_test,y_test)
```

```python
clf.score(X_test,y_test)
```

**Observations**:
- Using Random Forests we've got an average accuracy of 90.69%
