**Train Test Split of the Data**

```python
# Spliting the Data into Train and Test
from sklearn.model_selection import train_test_split
X = heart.drop(columns=['output'])
y = heart['output']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=40)
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
predicted = model.predict(X_test)
model.score(X_test, y_test)
```

```python

conf = confusion_matrix(y_test, predicted)
print ("Confusion Matrix : \n", conf)
```

**Accuracy of Logistic Regression**

```python
l_a=accuracy_score(y_test, predicted)*100
print ("The accuracy of Logistic Regression is : ", l_a, "%")
```

**Accuracy of Gaussian Naive Bayes**

```python
model = GaussianNB()
model.fit(X_train, y_train)
  
predicted = model.predict(X_test)
g_a=accuracy_score(y_test, predicted)*100
print("The accuracy of Gaussian Naive Bayes model is : ", g_a, "%")
```

**Accuracy of Support Vector Machine**

```python
model = SVC()
model.fit(X_train, y_train)
  
predicted = model.predict(X_test)
s_a=accuracy_score(y_test, predicted)*100
print("The accuracy of SVM is : ", s_a, "%")
```

**KNN**

```python
model = KNeighborsClassifier(n_neighbors = 1)  
model.fit(X_train, y_train)
predicted = model.predict(X_test)
  

print(confusion_matrix(y_test, predicted))
```

**Accuracy of KNN**

```python
knn_a=accuracy_score(y_test, predicted.round())*100
print("The accuracy of KNN is : ", knn_a, "%")
```

**Accuracy of Random forest**

```python
model = RandomForestRegressor(n_estimators = 100, random_state = 0)  
model.fit(X_train, y_train)  
predicted = model.predict(X_test)
r_a=accuracy_score(y_test, predicted.round())*100
print("The accuracy of Random Forest is : ", r_a, "%")
```
