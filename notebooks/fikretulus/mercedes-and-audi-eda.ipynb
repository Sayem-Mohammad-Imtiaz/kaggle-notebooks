##### In my country, we use km and liter instead of mileage and mpg. That's why I want to add this columns.

```python
m1=[]
for i in df["mileage"]:
    m=i*1.609344
    m1.append(m)
df["km"]=m1
df.head()
```

```python
L1=[]
for i in df["mpg"]:
    l=235.214583/i
    L1.append(l)
df["Consumption"]=L1
df.head()
```

```python
# We can do these visualizations by using for loop
columns=["model","year","transmission","fuelType"]

for i in columns:
    fig=px.histogram(df,x=df[i])
    fig.show()
```

```python
def Liter(Consumption):
    if (Consumption<5 and Consumption>0): return "Low"
    if (Consumption<9 and Consumption>=5): return "Medium"
    if (Consumption>=9): return "High"
df["Consume Rate"]=df.apply(lambda x: Liter(x["Consumption"]),axis=1)
sns.countplot(data=df,x="Consume Rate")
```

```python
plt.figure(figsize=(20,10))
sns.countplot(data=df,y="year")
```

```python
df_year=df["year"].value_counts().reset_index().rename(columns={"year":"count",
                                                               "index":"year"})
plt.figure(figsize=(12,6))
sns.lineplot(x="year",y="count",data=df_year)
```

```python
# Transmission and Price
df2=df.groupby("transmission")["price"].mean()
df2=pd.DataFrame(df2)
df2.reset_index()
sns.barplot(x="transmission",y="price",data=df2.reset_index())
```

```python
# Model and Price
df2=df.groupby("model")["price"].mean()
df2=pd.DataFrame(df2)
plt.figure(figsize=(20,20))
sns.barplot(x="model",y="price",data=df2.reset_index())
plt.xticks(rotation=90);
```

```python
#Most Expensive Car Models
df2.sort_values("price",ascending=False)[:15]
```
