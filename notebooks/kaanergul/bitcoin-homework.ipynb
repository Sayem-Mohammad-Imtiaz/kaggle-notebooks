```python
df = pd.read_csv("/kaggle/input/bitcoin-historical-data/bitstampUSD_1-min_data_2012-01-01_to_2021-03-31.csv")
df.head()
df.describe()   # informations about data set
df.columns      # columns name

```

```python
# Creating new data sets that represents each column seperately 

df.rename(columns = {"Volume_(BTC)":"Volume_BTC"},inplace = True)   # Removing paranthesis
df.rename(columns = {"Volume_(Currency)":"Volume_Currency"},inplace = True)   # Removing paranthesis

df_Timestamp= df.Timestamp
df_Open     = df.Open
df_High     = df.High
df_Low     = df.Low
df_Close     = df.Close
df_Volume_BTC     = df.Volume_BTC
df_Volume_Currency     = df.Volume_Currency
df_Weighted_Price     = df.Weighted_Price

```
