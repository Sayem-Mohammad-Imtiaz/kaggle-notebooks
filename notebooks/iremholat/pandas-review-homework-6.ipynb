**Indexing Pandas Time Series**

```python
# since we don't have dates in our data set, we have to add them
time_list = ['2000-11-21', '2000-01-22']
print(type(time_list[0]))   # as seen, the date is a string

# to convert it to a datetime object
datetime_object = pd.to_datetime(time_list)
print(type(datetime_object))
```

```python
# let's take the first 5 lines of the data and add it a time list
data2 = data.head()
date_list = ['2000-11-20', '2000-01-22', '2001-06-18', '2001-05-27', '2001-07-30']
datetime_object = pd.to_datetime(date_list)
data2['date'] = datetime_object

# let's change the indexes as dates
data2 = data2.set_index('date')
data2
```

```python
# now we can acces the specific lines according to the dates
print(data2.loc['2000-01-22'])
print(data2.loc['2001-06-01':'2001-07-30'])
```

**Resampling:** Statistical method over different time intervals. Needs string to specify frequency like 'M' for month and 'A' for year.

```python
data2.resample('A').mean()
```

```python
data2.resample('M').mean()
```

```python
# interpolate : filling the NaNs.
data2.resample('M').first().interpolate('linear')
```
