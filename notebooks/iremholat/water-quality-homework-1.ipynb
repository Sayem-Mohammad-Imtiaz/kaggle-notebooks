```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('../input/water-potability/water_potability.csv')
```

```python
data.corr() # to see if a variables is affected by the other variables
```

As seen above, none of the variables are significantly correlated with each other.

```python
# correlation map
f,ax = plt.subplots(figsize = (20,20))
sns.heatmap(data.corr(), annot = True, linewidths = .5, fmt = '.2f', ax = ax)
plt.show()
```

**Data Visualization With Matplotlib**

```python
# line plot of the variables Solids and Sulfates
data.Solids.plot(kind = 'line', color = 'pink', alpha = 1, label = 'Solids', linestyle = 'dotted', linewidth = 1, grid = True)
data.Sulfate.plot(kind = 'line', color = 'yellow', alpha = 1, label = 'Sulfate', linestyle = 'dashed', linewidth = 1, grid = True)
plt.legend('upper right')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Relationship Between the Solids and Sulfate in Water')
```

Both the line plot and the correlation shows that Solids and Sulfate are irrelevant.

```python
# scatter plot of Hardness and ph
data.plot(kind = 'scatter', color = 'lime', x = 'ph', y = 'Hardness', alpha = .2)
plt.title('The Relationship Between Water pH and Hardness')
plt.show()
```

```python
# histogram
data.Turbidity.plot(kind = 'hist', color = 'blue', alpha = .7, figsize = (6,6), bins = 10)
plt.show()
```

**Pandas**

```python
# Filtering pandas data frame
x = data['ph'] > 12
data[x]
```

```python
# Filtering with pandas Logical_and
data[np.logical_and(data['ph'] > 12, data['Conductivity'] < 450)]
```

```python
# Filtering using &
data[(data['ph']>12) & (data['Conductivity']<450)]
```
