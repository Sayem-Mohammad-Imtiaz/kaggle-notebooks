```python
df["item_id"] = df["item_id"].astype(int)
df["qty_ordered"] = df["qty_ordered"].astype(int)
df["Year"] = df["Year"].astype(int)
df["Month"] = df["Month"].astype(int)
```

## best selling category 
### - not considering item quantity

```python
df["category_name_1"].value_counts().plot.bar(figsize = (12,6) , title = "Best Selling categories")
plt.ylabel("Total Order")
plt.show()
```

## best selling category 
### -  considering item quantity

```python
df.groupby(['category_name_1'])['qty_ordered'].sum().sort_values(ascending=False)
```

```python
df.groupby(['category_name_1'])['qty_ordered'].sum().sort_values(ascending=False).plot.bar(figsize = (12,6) , title = "Best Selling categories")
plt.ylabel("Total Order with quantity")
plt.show()
```

```python
df.groupby(['Year','category_name_1'])['qty_ordered'].sum()
```

## best category in 2016

```python
df.groupby(['Year','category_name_1'])['qty_ordered'].sum()[2016].sort_values(ascending=False).plot.bar(figsize = (12,6) , title = "Best Selling categories")
plt.ylabel("Total Order with quantity")
plt.show()
```

## best category in 2017


```python
df.groupby(['Year','category_name_1'])['qty_ordered'].sum()[2017].sort_values(ascending=False).plot.bar(figsize = (12,6) , title = "Best Selling categories")
plt.ylabel("Total Order with quantity")
plt.show()
```

## best category in 2018


```python
df.groupby(['Year','category_name_1'])['qty_ordered'].sum()[2018].sort_values(ascending=False).plot.bar(figsize = (12,6) , title = "Best Selling categories")
plt.ylabel("Total Order with quantity")
plt.show()
```
