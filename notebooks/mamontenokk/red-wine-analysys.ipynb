```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
Y = df['alcohol']
```

```python
X = df.drop(['alcohol', 'quality'], axis=1)
```

```python
Y = np.array(Y)
```

```python
X = np.array(X)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state = 23)
```

```python
rgr = RandomForestRegressor(max_depth=50, random_state=23,
...                              n_estimators=100, bootstrap=True)
```

```python
rgr.fit(X_train, Y_train)
```

```python
rgr.score(X_train, Y_train)
```

```python
rgr.score(X_test, Y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree_regressor = DecisionTreeRegressor(max_depth = 100, random_state = 23, max_leaf_nodes=100)
```

```python
tree_regressor.fit(X_train, Y_train)
```

```python
tree_regressor.score(X_train, Y_train)
```

```python
tree_regressor.score(X_test, Y_test)
```
