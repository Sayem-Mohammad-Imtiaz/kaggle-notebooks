```python
import pandas as pd, seaborn as sb, numpy as np, matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../input/test-file/tested.csv")
data.head(7)
```

```python
data=data.drop(["Name","Cabin"],axis=1)
```

```python
data["Age"].fillna(data["Age"].median(),inplace=True)
```

```python
data["Fare"].fillna(data["Fare"].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data.Ticket
```

```python
x=data.drop(["Survived","Ticket"],axis=1)
y=data["Survived"]
```

```python
from sklearn.preprocessing import LabelEncoder
le_age = LabelEncoder()
le_embarked = LabelEncoder()

```

```python
x["Sex"]=le_age.fit_transform(data["Sex"])
x["Embarked"]=le_embarked.fit_transform(data["Embarked"])
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_predict)
```
