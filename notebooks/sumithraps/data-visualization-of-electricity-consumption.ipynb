```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as py
import plotly.offline
# date and time
import datetime as date

pd.set_option('display.max_columns', None)
pd.set_option('display.max_row', None)
import plotly as py
import plotly.graph_objs as go
from plotly.offline import iplot, init_notebook_mode
import bar_chart_race as bcr


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
# Import dataset
data=pd.read_csv('../input/state-wise-power-consumption-in-india/dataset_tk.csv')
long=pd.read_csv('../input/state-wise-power-consumption-in-india/long_data_.csv')
```

```python
# copy of Data ser
data_copy=data.copy()
```

```python
# Rename date
data.rename(columns={'Unnamed: 0':'date'}, inplace=True)
# change object into date and time formet 
data['date']=pd.to_datetime(data['date'])
long['Dates']=pd.to_datetime(long['Dates'])
data_copy=data.copy()   

# Reset index value 
data.set_index('date',inplace=True)
```

```python
# checking null values
data.isnull().sum()
```

```python
long.isnull().sum()
```

```python
bcr.bar_chart_race(data, figsize=(7, 4),period_length =500,filename = None, title='power usage by states')
```

```python
# data cleaning 
data['Total']=data.sum(axis=1)
data.loc['Total']=data.sum()
Total_comsumes=data.iloc[[-1]]
Total=Total_comsumes.transpose()
Total.drop(Total.index[[-1]], inplace=True)
Total.reset_index(inplace=True)
Total.rename(columns={'index':'State'}, inplace=True)
```

```python
# Bar chart of state wise electricity consumption
fig=go.Figure(go.Pie(labels=Total['State'], 
                     values=Total['Total'], title='State Wise Electricity Consumption'))
iplot(fig)
```

- In totaly Maharashtra state consume 12.7% power
- Gujarat  state consume 9.5% power 
- South states consume 5% to within 10% Electric city in india 


```python
# Top 10 Electricity consumption state 
Top_10_EC_Consume=Total.nlargest(10, 'Total')
Top_10_EC_Consume
```

```python
layout = go.Layout(title = "Maximum power consume state", xaxis = {'title':'EC'}, yaxis = {'title':'State'})
cons=go.Bar(x=Top_10_EC_Consume['Total'], y=Top_10_EC_Consume['State'], orientation='h' )
fig=go.Figure(data=cons, layout=layout)
iplot(fig)
print(Top_10_EC_Consume)
```

```python
# Minimum Power Consume state
Low_EC_Consumption=Total.nsmallest(10, 'Total')
```

```python
layout = go.Layout(title = "Minimum power consume state", xaxis = {'title':'EC'}, yaxis = {'title':'State'})
d=go.Bar(x=Low_EC_Consumption['Total'],y=Low_EC_Consumption['State'], orientation='h')
fig=go.Figure(data=d, layout=layout)
iplot(fig)
```

sikkim is very lowest ec comption state
