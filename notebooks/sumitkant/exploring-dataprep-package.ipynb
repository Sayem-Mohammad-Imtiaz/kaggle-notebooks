## 2. Load Dataset
using the credit card default dataset from UCI Machine Learning Repository. Before we pass this on to dataprep.ai, let's convert the numerically encoded categorical columns to category dtype, so they are not mistaken for numerical variables

```python
import pandas as pd
df = pd.read_csv('../input/default-of-credit-card-clients-dataset/UCI_Credit_Card.csv')
cat_vars = ['SEX', 'EDUCATION', 'MARRIAGE','PAY_0','PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6']
for v in cat_vars:
    df[v] = df[v].astype('category')
df.head()
```

## 3. Creating report
Creates a report with overview, descriptive statistics, histograms, correlations and missing values. It is as simple as printing hello world!

```python
from dataprep.eda import create_report
create_report(df)
```

## 4. Comparing Datasets
Can analyze difference between datasets by plotting distributions. Most likely you are comparing in-sample with out-of-sample or out-of-time datasets. This is a handy feature. The problem clearly is the that it does not normalize by the size of the dataset and the differences look stark because of the frequency of values in the datasets

```python
# train test split
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size = 0.3)
print(df.shape, train.shape, test.shape)

# plotting diff
from dataprep.eda import plot_diff
plot_diff([train, test])
```
