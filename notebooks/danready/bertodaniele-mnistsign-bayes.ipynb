Ho provato ad applicare alcuni algoritmi Naive Bayes presenti in sklearn.naive_nayes al dataset MNIST sign-language.

```python
import numpy as np
import pandas as pd
from sklearn import naive_bayes
from sklearn import metrics
```

```python
train_data = pd.read_csv("../input/sign-language-mnist/sign_mnist_train/sign_mnist_train.csv") #reading the csv files using pandas
test_data = pd.read_csv("../input/sign-language-mnist/sign_mnist_test/sign_mnist_test.csv")
```

```python
#Preparing data

y_train = train_data['label']

X_train = train_data.drop(columns = 'label')

y_test = test_data['label']

X_test = test_data.drop(columns = 'label')
```

```python
# GaussianNB model
model = naive_bayes.GaussianNB()
model.fit(X_train, y_train)
# predict
y_pred_1 = model.predict(X_test)
# accuracy
print("GaussianNB accuracy:", metrics.accuracy_score(y_test, y_pred_1), "\n")
```

```python
# MultinomialNB model
model = naive_bayes.MultinomialNB()
model.fit(X_train, y_train)
# predict
y_pred_2 = model.predict(X_test)
# accuracy
print("MultinomialNB accuracy:", metrics.accuracy_score(y_test, y_pred_2), "\n")
```

```python
# ComplementNB model
model = naive_bayes.ComplementNB()
model.fit(X_train, y_train)
# predict
y_pred_3 = model.predict(X_test)
# accuracy
print("ComplementNB accuracy:", metrics.accuracy_score(y_test, y_pred_3), "\n")
```

```python
print("GaussianNB accuracy:", metrics.accuracy_score(y_test, y_pred_1), "\n")
print("MultinomialNB accuracy:", metrics.accuracy_score(y_test, y_pred_2), "\n")
print("ComplementNB accuracy:", metrics.accuracy_score(y_test, y_pred_3), "\n")
```
