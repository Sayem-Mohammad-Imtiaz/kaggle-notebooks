Nell'ambito delle SVM, tentativo di applicare un semplice modello lineare e uno non lineare con iperparametri di default al dataset mnist relativo al linguaggio dei segni.

Vengono utilizzati gli strumenti forniti da sklearn (SVC di sklearn.svm)

```python
import numpy as np
import pandas as pd
from sklearn.svm import SVC
```

```python
train_data = pd.read_csv("../input/sign-language-mnist/sign_mnist_train/sign_mnist_train.csv") #reading the csv files using pandas
test_data = pd.read_csv("../input/sign-language-mnist/sign_mnist_test/sign_mnist_test.csv")
```

```python
#Preparing data

y_train = train_data['label']

X_train = train_data.drop(columns = 'label')

#Normalization
X_train = X_train/255.0

#scaling the features
from sklearn.preprocessing import scale
X_train = scale(X_train)

y_test = test_data['label']

X_test = test_data.drop(columns = 'label')

X_test = X_test/255.0

X_test = scale(X_test)
```

```python
# linear model

model_linear = SVC(kernel='linear')
model_linear.fit(X_train, y_train)
```

```python
# predict
y_pred = model_linear.predict(X_test)
```

```python
from sklearn import metrics

# accuracy
print("accuracy:", metrics.accuracy_score(y_test, y_pred), "\n")
```

```python
# non-linear model
non_linear_model = SVC(kernel='rbf')

# fit
non_linear_model.fit(X_train, y_train)

# predict
y_pred = non_linear_model.predict(X_test)
```

```python
#accuracy
print("accuracy:", metrics.accuracy_score(y_test, y_pred), "\n")
```
