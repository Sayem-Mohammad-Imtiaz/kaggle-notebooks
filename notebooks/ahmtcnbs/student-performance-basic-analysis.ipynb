```python
df['OverAllPassStatus'] = df.apply(lambda x:'F' 
                                   if x['MathPassStatus'] == 'F' 
                                   or x['WritingPassStatus'] == 'F'
                                   or x['ReadingPassStatus'] == 'F'
                                   else 'P', axis=1)
df.OverAllPassStatus.value_counts()
```

```python
df['TotalMarks'] = df['MathScore'] + df['WritingScore'] + df['WritingScore']
df['Mean'] = df['TotalMarks']/3
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(df['sex'])
df['NewSex'] = le.transform(df['sex'])
df.head()
```

```python
def GetGrade(Mean,OverAllPassStatus):
    if (OverAllPassStatus == 'F'):
        return 'F'
    if (Mean >= 80):
        return 'A'
    if (Mean >= 70):
        return 'B'
    if (Mean >= 60):
        return 'C'
    if (Mean >= 50):
        return 'D'
    if (Mean >= 40):
        return 'E'
    else: 
        return 'F'
    
df['GradeStatus'] = df.apply(lambda x: GetGrade(x['Mean'],x['OverAllPassStatus']),axis=1)

df.GradeStatus.value_counts().sort_values(ascending=False)
```

```python
sns.countplot(data=df, x='GradeStatus',order=['A','B','C','D','E','F'],palette='muted')
```

```python
p = sns.countplot(x='parental level of education',data=df, hue='GradeStatus',palette='muted')
plt.setp(p.get_xticklabels(),rotation=90)
```
