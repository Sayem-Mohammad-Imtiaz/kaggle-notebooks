```python
df = pd.read_csv('/kaggle/input/pima-indians-diabetes-database/diabetes.csv')
df.head()
```

```python
df.info()
```

```python
y = df["Outcome"]
x = df.drop(['Outcome'] ,axis = 1)
```

```python
x.head()
```

```python
y.head()
```

```python
len(y)
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test ,y_train ,y_test = train_test_split(x ,y ,test_size = 0.2 ,random_state = 42)
x_train.shape ,x_test.shape ,y_train.shape ,y_test.shape
```

## 1-Stochastic Gradient Decsent

```python
from sklearn.linear_model import SGDClassifier
sgd = SGDClassifier()
sgd.fit(x_train ,y_train)
```

## K-fold Cross Validation

```python
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(sgd ,x_train ,y_train ,cv = 3 ,scoring = 'accuracy')
accuracies
```

Let's try to scale the inputs..

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
cross_val_score(sgd ,x_train_scaled ,y_train ,cv = 3 ,scoring = 'accuracy')
```

the accuracies are now slightly better ..

## some error analysis

```python
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_predict
y_predict = cross_val_predict(sgd ,x_train_scaled ,y_train ,cv = 3)
conf_mat = confusion_matrix(y_train ,y_predict)
```

```python
conf_mat
```

```python
import matplotlib.pyplot as plt
plt.matshow(conf_mat, cmap=plt.cm.gray)
plt.show()
```

##  rows represent actual classes, while columns represent predicted classes

means that there are (306) examples are correctly classified as non diabetic patient ,and (95) are wrongly classified as diabetic patient ,while (81) are wrongly classified as non-diabetic and (132) correctly classified as diabetic //

## Now let's test on the test data

```python
import numpy as np
x_test_scaled = scaler.transform(x_test)
test_predict = sgd.predict(x_test_scaled)
```

```python
test_acc = np.mean(test_predict==y_test)
test_acc
```

## 2-Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(solver = 'liblinear')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
log_reg.fit(x_train_scaled ,y_train)
```

```python
from sklearn.model_selection import cross_val_score
train_accuracies = cross_val_score(log_reg ,x_train_scaled ,y_train ,cv = 3 ,scoring = 'accuracy')
train_accuracies
```

```python
x_test_scaled = scaler.transform(x_test)
test_predictions = log_reg.predict(x_test_scaled)
```

```python
import numpy as np
train_predictions = log_reg.predict(x_train_scaled)
train_accuracy = np.mean(train_predictions == y_train)
train_accuracy
```
