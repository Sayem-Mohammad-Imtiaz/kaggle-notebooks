```python
import pandas as pd
a=pd.read_csv("../input/gufhtugu-publications-dataset-challenge/GP Orders - 2.csv",encoding = "cp1252")
b=pd.read_csv("../input/gufhtugu-publications-dataset-challenge/GP Orders - 4.csv")
data=pd.concat([a,b],axis=0)
data.head(10000)
#print(len(a),len(data))
```

```python
import numpy as np

def Remove_dump_values(data, cols):
    for col in cols:
        data[col] = np.where(data[col] == '-', 'None', data[col])
    return data
cols = data.columns
data_bin = Remove_dump_values(data, cols)
data_bin=data_bin[:38374]
Y=data_bin['Book Name']
data_bin=data_bin[(data_bin['Order Status'] == "Completed")]
```

**Part 1**

```python
books=data_bin['Book Name'].unique() 
books=pd.DataFrame(books)
books
```

```python
counts=data_bin['Book Name'].value_counts()
#counts=pd.DataFrame(counts)
#final_data=pd.concat([data_bin,counts],axis=1)
#final_data
print(len(counts),len(books))
```

```python
count=[]
for i in range(len(counts)):
    count.append(counts[i])
count=pd.DataFrame(count)
books.columns={'books'}
count.columns={'count'}
final_data=pd.concat([books,count],axis=1)
```

```python
final_data.columns
```

```python
str(final_data['books'][0])
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(200,100))
sns.set(font_scale=10)
sns.set(style='darkgrid')
sns.set_context(font_scale=0.5)
x_axis1=[]
for i in range(0,10):
  x_axis1.append(str(final_data['books'][i]))
x_axis=[]
for i in range(0,10):
  x_axis.append(i)

plott = sns.barplot(x='books',y='count',data=final_data[0:10])
plt.xlabel("Top Ten Selling Books in Pakistan",size=250)
plt.ylabel("Frequency",size=250)
plt.xticks(np.arange(0,10),x_axis1,rotation=90,fontsize=180,linespacing=150)
plt.yticks(fontsize=180)
plt.show()
```

**Cities that buy the most books**

```python
city=data_bin['City (Billing)'].unique() 
counts=data_bin['City (Billing)'].value_counts()
print(len(counts),len(city))
count=[]
for i in range(len(counts)):
    count.append(counts[i])
count=pd.DataFrame(count)
city=pd.DataFrame(city)
city.columns={'city'}
count.columns={'count'}
final_data=pd.concat([city,count],axis=1)
#final_data.columns={'0','City'}
str(final_data['city'][0])
```

```python
str(final_data['city'][0])
```

```python

import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(200,100))
sns.set(font_scale=10)
sns.set(style='darkgrid')
sns.set_context(font_scale=0.5)
x_axis1=[]
for i in range(0,10):
  x_axis1.append(str(final_data['city'][i]))
x_axis=[]
for i in range(0,10):
  x_axis.append(i)

plott = sns.barplot(x='city',y='count',data=final_data[0:10])
plt.xlabel("Top Cities",size=250)
plt.ylabel("Frequency",size=250)
plt.xticks(np.arange(0,10),x_axis1,rotation=90,fontsize=180,linespacing=150)
plt.yticks(fontsize=180)
plt.show()
```
