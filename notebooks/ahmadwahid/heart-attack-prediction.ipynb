### Importing the required Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , mean_absolute_error
```

```python
# Importing dataset i.e heart.csv
ht_data=pd.read_csv('../input/heart-attack-analysis-prediction-dataset/heart.csv')
ht_data.head()
```

```python
# Importing second dataset i.e o2Saturation.csv
st_data=pd.read_csv('../input/heart-attack-analysis-prediction-dataset/o2Saturation.csv')
st_data.head()
```

```python
# To know if there is any null value in both datasets
print(ht_data.isnull().sum())
print(st_data.isnull().sum())
```

### Visualizing the Data

```python
# Visualize heart.csv by using pairplot ft of seaborn
sns.pairplot(data=ht_data)
```

```python
# Preparing the data
X=ht_data.drop('output',axis=1)
y=ht_data['output']
X.head()
```

```python
# Spliting the data into training and test sets using train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

### Training the Model

```python
#  Training Algorithm
regressor=LogisticRegression().fit(X_train,y_train)
predicted_output=regressor.predict(X_test)
predicted_output
```

```python
# Predicting the score of the model 
print('The accuracy score of the model is {}'.format(accuracy_score(y_test,predicted_output)))
```

### Evaluating the Model

```python
print('Mean Absolute Error : ',mean_absolute_error(y_test,predicted_output))
```
