Parameter info

```python
df.info()
```

* Total features: 14
* There are 2 data types: int64 & float64
* No "Null-value" in the data

* Age: Age of patient
* Sex : Sex of the patient (1:male, 0:female)
* exang: exercise induced angina (1 = yes; 0 = no)
* ca: number of major vessels (0-3)
* cp : Chest Pain type chest pain type
    (Value 1: typical angina,
    Value 2: atypical angina,
    Value 3: non-anginal pain,
    Value 4: asymptomatic)
* trtbps : resting blood pressure (in mm Hg)
* chol : cholestoral in mg/dl fetched via BMI sensor
* fbs : (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
* oldpeak - ST depression induced by exercise relative to rest
* rest_ecg : resting electrocardiographic results
* slp: slope - the slope of the peak exercise ST segment (2 = upsloping; 1 = flat; 0 = downsloping)
* thalach : maximum heart rate achieved
* thal :2 = normal; 1 = fixed defect; 3 = reversable defect
* target : 0= less chance of heart attack, 1= more chance of heart attack

Load libraries for modeling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df_copy = df
```

```python
cat_col=["sex","exng","caa","cp","fbs","restecg","slp","thall"]
num_col=["age","trtbps","chol","thalachh","oldpeak"]

df_copy = pd.get_dummies(df_copy, columns = cat_col, drop_first = True)
X = df_copy.drop(['output'],axis=1)
y = df_copy[['output']]
```

```python
scaler = RobustScaler()
X[num_col] = scaler.fit_transform(X[num_col])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

y_pred_proba = logreg.predict_proba(X_test)
y_pred = np.argmax(y_pred_proba,axis=1)

print("The test accuracy: ", accuracy_score(y_test, y_pred))
```
