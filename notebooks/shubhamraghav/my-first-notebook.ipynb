```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("../input/bangalore-real-estate-price/blr_real_estate_prices.csv")
df.head(3)
```

```python
df2 = df.drop(['area_type','society','balcony','availability'],axis='columns')
df2
```

```python
df2.isnull().sum()
```

```python
df3=df2.dropna()
df3.isnull().sum()
```

```python
df3.shape
```

```python
df3['size'].unique()
```

```python
dummies = pd.get_dummies(df3.size)
dummies
```

```python
X = df3.drop('price',axis='columns')
Y = df3.price

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=10)

from sklearn.linear_model import LinearRegression
hello = LinearRegression()
hello.fit(X_train, Y_train)
hello.score(X_test, Y_test)
```
