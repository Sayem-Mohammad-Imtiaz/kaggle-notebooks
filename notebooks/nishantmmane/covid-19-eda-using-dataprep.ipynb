```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns 
import matplotlib.pyplot as plt
import warnings
from wordcloud import WordCloud
warnings.filterwarnings("ignore")
```

```python
#pip install wordcloud
df=pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df.total_vaccinations.fillna(0,inplace=True)	
df.people_vaccinated.fillna(0,inplace=True)	
df.people_fully_vaccinated.fillna(0,inplace=True)	
df.daily_vaccinations_raw.fillna(0,inplace=True)	
df.daily_vaccinations.fillna(0,inplace=True)	
df.total_vaccinations_per_hundred.fillna(0,inplace=True)	
df.people_vaccinated_per_hundred.fillna(0,inplace=True)	
df.people_fully_vaccinated_per_hundred.fillna(0,inplace=True)	
df.daily_vaccinations_per_million.fillna(0,inplace=True)	
df.dropna(inplace=True)
df.head()
```

```python
new_df = df.groupby(["country",'iso_code','vaccines'])['total_vaccinations','people_vaccinated',
'people_fully_vaccinated','daily_vaccinations','total_vaccinations_per_hundred','people_vaccinated_per_hundred',
            "people_fully_vaccinated_per_hundred",'daily_vaccinations_per_million'].max().reset_index()
new_df.head()
```

```python
month = [date.split('-')[1] for date in df.date]
df['month'] = month
```

**total vaccinations country wise**

```python
plt.figure(figsize=(14,7))
df.groupby('country')['total_vaccinations'].max().sort_values(ascending=False)[:30].plot(kind="bar")
plt.show()
```

**People fully vaccined by country**

```python
plt.figure(figsize=(14,7))
df.groupby('country')['people_fully_vaccinated'].max().sort_values(ascending=False)[:30].plot(kind="bar")
```

**daily vaccinations by countrywise**

```python
plt.figure(figsize=(14,7))
df.groupby('country')['daily_vaccinations'].max().sort_values(ascending=False)[:30].plot(kind="bar")
plt.show()
```

**total vaccinations per hundred by countrywise**

```python
plt.figure(figsize=(14,7))
df.groupby('country')['total_vaccinations_per_hundred'].max().sort_values(ascending=False)[:30].plot(kind="bar")
plt.show()
```

*** over all total vaccinations ***

```python
plt.figure(figsize=(14,7))
df.groupby('month')['daily_vaccinations'].sum().sort_values(ascending=True)[:30].plot(kind="bar")
plt.show()
```
