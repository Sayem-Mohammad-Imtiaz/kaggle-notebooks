# **Import packages**

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import metrics
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler
import xgboost as xgb
acc=[]
model=[]
```

```python
sc = RobustScaler()
x = df.iloc[:, :-1].values
y = df.iloc[:, -1].values

xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2,random_state=0)

xtrain = sc.fit_transform(xtrain)
xtest = sc.transform(xtest)

print("xtrain shape:= ",xtrain.shape)
print("xtest shape:= ",xtest.shape)
print("ytrain shape:= ",ytrain.shape)
print("xtest shape:= ",ytest.shape)
```

```python
### Decision Tree
DecisionTree = DecisionTreeClassifier(criterion="entropy",random_state=2,max_depth=5)
DecisionTree.fit(xtrain,ytrain)
pred_value=DecisionTree.predict(xtest)
x=metrics.accuracy_score(ytest,pred_value)
acc.append(x)
model.append('Decision Tree')
print("DecisionTrees's Accuracy is: ", x*100)
print(classification_report(ytest,pred_value))
```

**Guassian Naive Bayes**

```python
NB=GaussianNB()
NB.fit(xtrain,ytrain)
pred_value=NB.predict(xtest)
x=metrics.accuracy_score(ytest,pred_value)
acc.append(x)
model.append('Guassian Naive Bayes')
print("Guassian Naive Bayes Accuracy is: ", x*100)
print(classification_report(ytest,pred_value))
```

**SVM**

```python
svm=SVC()
svm.fit(xtrain,ytrain)
pred_value=svm.predict(xtest)
x=metrics.accuracy_score(ytest,pred_value)
acc.append(x)
model.append('SVM')
print('SVM accuracy is: ',x*100)
print(classification_report(ytest,pred_value))
```

**Logistic Regression**

```python
logisreg=LogisticRegression(random_state=2)
logisreg.fit(xtrain,ytrain)
pred_value=logisreg.predict(xtest)
x=metrics.accuracy_score(ytest,pred_value)
acc.append(x)
model.append('Logistic Regression')
print('Logistic Regression: ',x*100)
print(classification_report(ytest,pred_value))
```

**RandomForest**

```python
RF = RandomForestClassifier(n_estimators=20, random_state=0)
RF.fit(xtrain,ytrain)

pred_value = RF.predict(xtest)

x = metrics.accuracy_score(ytest, pred_value)
acc.append(x)
model.append('RF')
print("RF's Accuracy is: ", x)

print(classification_report(ytest,pred_value))
```

**XG Boost**

```python
XB = xgb.XGBClassifier()
XB.fit(xtrain,ytrain)

pred_value = XB.predict(xtest)

x = metrics.accuracy_score(ytest, pred_value)
acc.append(x)
model.append('XGBoost')
print("XGBoost's Accuracy is: ", x)

print(classification_report(ytest,pred_value))
```

```python
plt.figure(figsize=[10,5],dpi = 100)
plt.title('Accuracy Comparison')
plt.xlabel('Accuracy')
plt.ylabel('Algorithm')
sns.barplot(x = acc,y = model,palette='dark')
```

**SVM Hypertunning**

```python
svc = SVC(kernel = 'rbf',C = 1, degree = 5, random_state = 0)
svc.fit(xtrain,ytrain)
pred = svc.predict(xtest)
x=metrics.accuracy_score(pred,ytest)
print("Accuracy is:= ",x)
print(classification_report(ytest,pred_value))
```
