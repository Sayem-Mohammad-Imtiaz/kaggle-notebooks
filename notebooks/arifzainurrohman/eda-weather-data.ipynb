```python
data = data.set_index('Formatted Date')
data.head()
```

```python
#after resampling

data_columns = ['Wind Speed (km/h)','Temperature (C)']
df_monthly_mean = data[data_columns].resample('MS').mean()
df_monthly_mean.head()
```

```python
#Plotting Variation in Apparent Wind Speed (km/h) and Temperature (C) with time
import numpy as np             
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

plt.figure(figsize=(14,6))
plt.title("Variation in Apparent Wind Speed (km/h) and Temperature (C) with time")
sns.lineplot(data=df_monthly_mean)
```

Observation: From the plot above, we can say that wind speed will increase when the air temperature drops. And vice versa.

```python
data_precip = data[['Precip Type','Temperature (C)']]
data_precip
```

```python
data_precip = data_precip.groupby(['Precip Type']).mean()[['Temperature (C)']]
data_precip
```

Observation : From the plot above, we can say that the air temperature will drop when the weather is rainy and snowy. 

```python
# data to plot
n_groups = 2

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.25
opacity = 0.9

rects1 = plt.bar(index, data_precip['Temperature (C)'], bar_width,
alpha=opacity,
color='b',
label='Temperature (C)')


plt.rcParams["figure.figsize"] = (4,2)
plt.xlabel('Weather')
plt.ylabel('Visibility & Wind Speed')
plt.title('Variation in Apparent Visibility and Wind Speed with Summary')
plt.xticks(index,('rain','snow'))
plt.legend()
plt.show()
```

```python
data_1 = pd.read_csv('/kaggle/input/weather-dataset/weatherHistory.csv')
data_visibility = data_1.groupby(['Summary']).mean()[['Visibility (km)','Wind Speed (km/h)']]
visi = data_visibility.sort_values(by=['Visibility (km)'],ascending=False)
visi = visi.head()
visi
```

```python
# data to plot
n_groups = 5

# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.25
opacity = 0.9

rects1 = plt.bar(index, visi['Visibility (km)'], bar_width,
alpha=opacity,
color='r',
label='Visibility (km)')

rects2 = plt.bar(index + bar_width, visi['Wind Speed (km/h)'], bar_width,
alpha=opacity,
color='g',
label='Wind Speed (km/h)')

plt.rcParams["figure.figsize"] = (16,5)
plt.xlabel('Weather')
plt.ylabel('Visibility & Wind Speed')
plt.title('Variation in Apparent Visibility and Wind Speed with Summary')
plt.xticks(index+0.12,('Partly Cloudy','Windy and Partly Cloudy','Breezy and Mostly Cloudy','Dangerously Windy and Partly Cloudy','Clear'))
plt.legend()
plt.show()
```
