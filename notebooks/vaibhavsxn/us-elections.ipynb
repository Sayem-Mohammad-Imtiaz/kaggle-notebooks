### Taking a sneak peak into the files

```python
county_candidate = pd.read_csv('/kaggle/input/us-election-2020/senate_county_candidate.csv')
```

```python
county_candidate.shape
```

```python
county_candidate.groupby('candidate').sum().sort_values(by = 'total_votes', ascending = False).head(20).plot(kind = 'bar', figsize = (60, 36));
```

```python
county_candidate.groupby('candidate').sum().sort_values(by = 'total_votes', ascending = False).head(20)
```

### Lets see the winning the party
