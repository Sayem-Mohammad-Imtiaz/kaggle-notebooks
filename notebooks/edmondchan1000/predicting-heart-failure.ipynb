```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# read in the data
df = pd.read_csv('../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv')
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop('DEATH_EVENT', axis=1)
y = df['DEATH_EVENT']
sc = StandardScaler()
X = sc.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Try Logistic Regression 

```python
# Logistic Regression

from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr_predictions = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, lr_predictions))
print('\n')
print(confusion_matrix(y_test, lr_predictions))
```

# Trying KNeighborsClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(X_train, y_train)
```

```python
knn_preds = knn.predict(X_test)
```

```python
print(classification_report(y_test, knn_preds))
print('\n')
print(confusion_matrix(y_test, knn_preds))
```

```python
# elbow method to improve knn model score
# code from Jose Portilla
error_rate = []

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
# code from Jose Portilla
plt.figure(figsize = (10,6))
plt.plot(range(1,40), error_rate)
```

```python
# 29 has the lowest error rate in the graph above. Use that for n_neighbors parameter.
knn = KNeighborsClassifier(n_neighbors = 29)
knn.fit(X_train, y_train)
knn_preds = knn.predict(X_test)

print(classification_report(y_test, knn_preds))
print('\n')
print(confusion_matrix(y_test, knn_preds))
```

# Trying Support Vector Models

```python
from sklearn.svm import SVC
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

```python
svc_preds = svc_model.predict(X_test)
```

```python
print(classification_report(y_test, svc_preds))
print('\n')
print(confusion_matrix(y_test, svc_preds))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# code from Jose Portilla
param_grid = {'C': [0.1,1,10,100], 'gamma': [1,0.1,0.01,0.001]}
```

```python
# code from Jose Portilla
grid = GridSearchCV(SVC(), param_grid, verbose=2)
grid.fit(X_train, y_train)
```

```python
grid_preds = grid.predict(X_test)
```

```python
print(classification_report(y_test, grid_preds))
print('\n')
print(confusion_matrix(y_test, grid_preds))
```
