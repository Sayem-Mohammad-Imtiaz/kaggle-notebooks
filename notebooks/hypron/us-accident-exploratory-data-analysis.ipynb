The accident data are collected from February 2016 to Dec 2020, using two APIs that provide streaming traffic incident (or event) data. These APIs broadcast traffic data captured by a variety of entities, such as the US and state departments of transportation, law enforcement agencies, traffic cameras, and traffic sensors within the road-networks. Currently, there are about 4.2 million accident records in this dataset.

## When Most Accident Occurs During Day

# Start Time

```python
df.Start_Time
```

* Figure out how to show percentage

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
sns.distplot(df.Start_Time.dt.hour, bins=24, kde=False, norm_hist=True)
```

So As We Can See Most Accident Occurs in the Morning Between 6am to 10 Am. As people leave to their Work in hurry.
* Next Highest Percentage is between 3pm to 6pm

* Is the Distributed Data of accidents by hour the same on weekends as on weekdays

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins=7, kde=False, norm_hist=True)
```

```python
sundays_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sundays_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

* Let's Check Out Data About Monday

```python
monday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(monday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

* So As We Can See Sundays Occurs Most Accident Between 10 Am to 3 Pm. Unlike Weekdays

```python
df_2019 = df[df.Start_Time.dt.year == 2019]
df_2019_Bing = df_2019[df_2019.Source == 'MapQuest']
sns.distplot(df_2019_Bing.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```

* Lets Check Out Monthly-Wise data of accidents?

* Some Data is Missing between 2016 to 2017.
