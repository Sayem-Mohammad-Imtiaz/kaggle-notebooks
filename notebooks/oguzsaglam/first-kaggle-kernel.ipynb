Reading csv file

```python
# df = data frame
df = pd.read_csv("/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv")  # read csv file
# df.head()
```

```python
print(df.columns)  # receive columns so that we can reach data easily.
```

```python
latest_date = '2021-01-29'  # specify the date
date_frame = df['date'] == latest_date  # get the latest date frames.

# print(df[date_frame].total_vaccinations)  # debug total vaccinations in 29.01.2021
# print(pd.factorize(df['iso_code'])[0])  # convert iso codes into integer id's (categorical conversation)

# create a new column called categoric iso 
df['categoric_iso'] = df['iso_code'].astype('category')  # since the datatype of the iso_code column is string, convert it to categoric data
df['categoric_iso'] = pd.factorize(df['categoric_iso'])[0]  # convert categoric data to integers, just like id's in sql.

# more dubugging
# print(df['categoric_iso'])
# for row in df['categoric_iso']:
#    print(row)
# print(df.columns)
print(df[date_frame].categoric_iso)
print(df[date_frame].total_vaccinations)
```

Plot the data

```python
# plot the outcome, which the data is total vaccinations for 59 countries in 29/01/2021.
df[date_frame].categoric_iso.plot(kind = 'line', color = 'g',label = 'country',linewidth=3,alpha = 0.7,grid = True,linestyle = ':')
df[date_frame].total_vaccinations.plot(color = 'b',label = 'total vacs',linewidth=2, alpha = 0.7,grid = True,linestyle = '-.')
import matplotlib.pyplot as plt
plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('country iso codes')   # label = name of label
plt.ylabel('total vaccinations')
plt.title('Line Plot')            # title = title of plot
plt.show()
```
