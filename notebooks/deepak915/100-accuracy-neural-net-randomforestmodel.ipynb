# **Creating dummy variable for each category**

```python
Dummies=pd.get_dummies(data[['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat']])
```

```python
Dummies.head()
```

```python
Dummies.shape
```

# **Converting target class to numeric**

```python
y=data['class'].replace({'e':0,'p':1})
```

```python
y.head()
```

```python
y.shape
```

# Scaling features

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(Dummies)
```

```python
x
```

# **Spliting train and test dataset**

```python
from sklearn.model_selection import train_test_split
xtr,xte,ytr,yte=train_test_split(x,y,test_size=0.3,random_state=42)
```

# **Initializing model**

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=150,criterion='entropy')
rfc.fit(xtr,ytr)
```

# **Predicting data**

```python
from sklearn.metrics import classification_report
```

```python
ypte=rfc.predict(xte)
```

```python
cr=print(classification_report(yte,ypte))
```

# **We've got a 100% accuracy!**

```python
x.shape
```

```python
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=100,activation='sigmoid',input_shape=(117,)))
model.add(tf.keras.layers.Dense(units=200, activation='sigmoid'))
model.add(tf.keras.layers.Dense(units=300, activation='sigmoid'))
model.add(tf.keras.layers.Dense(units=200, activation='sigmoid'))
model.add(tf.keras.layers.Dense(units=1, activation='linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='mean_squared_error')
```

```python
hist=model.fit(xtr,ytr,epochs=100,batch_size= 50, validation_split=0.23)
```

```python
hist.history.keys()
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```

```python
ypr=model.predict(xte)
plt.plot(ypr, yte)
```
