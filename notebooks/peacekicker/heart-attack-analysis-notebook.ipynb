```python
index = df[(df['age']>71) | (df['age']< 34)].index
df.drop(index,inplace=True)
df.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
xtrain,xtest,ytrain,ytest = train_test_split(x,y,random_state=42,test_size=.2)
```

```python
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
```

```python
lr= LogisticRegression()
lr.fit(xtrain,ytrain)
print("Train:",lr.score(xtrain,ytrain))
print("Test:",lr.score(xtest,ytest))
```

```python
dc=DecisionTreeClassifier(max_depth=13)
dc.fit(xtrain,ytrain)
print("Train:",dc.score(xtrain,ytrain))
print("Test:",dc.score(xtest,ytest))
```

```python
svc=SVC()
svc.fit(xtrain,ytrain)
print("Train:",svc.score(xtrain,ytrain))
print("Test:",svc.score(xtest,ytest))
```

```python
rf=RandomForestClassifier(n_estimators=250)
rf.fit(xtrain,ytrain)
print("Train:",rf.score(xtrain,ytrain))
print("Test:",rf.score(xtest,ytest))
```

```python
knn=KNeighborsClassifier()
knn.fit(xtrain,ytrain)
print("Train:",knn.score(xtrain,ytrain))
print("Test:",knn.score(xtest,ytest))
```
