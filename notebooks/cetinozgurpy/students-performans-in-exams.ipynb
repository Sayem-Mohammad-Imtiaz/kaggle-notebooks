>  **Succes rate between boys and girls**


```python
df = pd.read_csv("../input/students-performance-in-exams/StudentsPerformance.csv")

df_girls = df[df.gender=="female" ]
count_girls = len(df_girls[df_girls["math score"] > 90])

df_boys = df[df.gender =="male"]
count_boys = len(df_boys[df_boys["math score"] > 90])

print(count_boys,count_girls)
```

As we can see, here i sorted and showed the boys and girls which have over 90 points in math
