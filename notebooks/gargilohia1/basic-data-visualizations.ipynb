```python
df = pd.read_csv("/kaggle/input/students-performance-in-exams/StudentsPerformance.csv")
```

```python
for col in df.columns :
    print(col)
```

```python
#Scatter plot for writing scores and reading scores
sns.scatterplot(data=df, x="reading score", y="writing score",hue="race/ethnicity")
```

```python
sns.relplot(data=df, x="reading score", y="writing score", hue="gender")
```

```python
a = df.groupby('race/ethnicity')['parental level of education'].count()
```

```python
df_new = df.groupby('race/ethnicity').size()

df_new.plot(kind='pie', subplots=True)
   
plt.title("Pie Chart of Race/ethnicity")
plt.ylabel("")
plt.show()
```

```python
sns.set_theme(style="whitegrid")
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)   
ax= sns.boxplot(data=df, x="race/ethnicity", y="math score",ax=ax)
```

```python
g = sns.catplot(x="race/ethnicity", y="math score",
                hue="lunch",
                data=df, kind="box");
```

```python
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)    
sns.stripplot(x="race/ethnicity", y="reading score", data=df, ax=ax)
plt.show()
```

```python
plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df, kind="scatter", hue="lunch", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5))
plt.show()
```

```python
plt.figure(figsize=(10,8), dpi= 80)
sns.pairplot(df, kind="scatter", hue="gender", plot_kws=dict(s=80, edgecolor="white", linewidth=2.5))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)   
sns.kdeplot(data=df, x="reading score", hue="race/ethnicity", multiple="fill",ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)   
sns.kdeplot(
    data=df, x="reading score", y="writing score",
    fill=True, thresh=0, levels=100, cmap="mako", ax=ax
)
```
