```python
#summarizing data by Jungian functions 
se_dom = df[df.type.eq('ESTP')].append(df[df.type.eq('ESFP')])
si_dom = df[df.type.eq('ISFJ')].append(df[df.type.eq('ISTJ')])
ne_dom = df[df.type.eq('ENFP')].append(df[df.type.eq('ENTP')])
ni_dom = df[df.type.eq('INTJ')].append(df[df.type.eq('INFJ')])
fe_dom = df[df.type.eq('ENFJ')].append(df[df.type.eq('ESFJ')])
fi_dom = df[df.type.eq('ISFP')].append(df[df.type.eq('INFP')])
te_dom = df[df.type.eq('ENTJ')].append(df[df.type.eq('ESTJ')])
ti_dom = df[df.type.eq('INTP')].append(df[df.type.eq('ISTP')])

print('Se dom count: ' + str(se_dom.posts.count()) + ' - ' + str(sum(map(len, se_dom.posts))) + ' characters')
print('Si dom count: ' + str(si_dom.posts.count()) + ' - ' + str(sum(map(len, si_dom.posts))) + ' characters')
print('Ne dom count: ' + str(ne_dom.posts.count()) + ' - ' + str(sum(map(len, ne_dom.posts))) + ' characters')
print('Ni dom count: ' + str(ni_dom.posts.count()) + ' - ' + str(sum(map(len, ni_dom.posts))) + ' characters')
print('Fe dom count: ' + str(fe_dom.posts.count()) + ' - ' + str(sum(map(len, fe_dom.posts))) + ' characters')
print('Fi dom count: ' + str(fi_dom.posts.count()) + ' - ' + str(sum(map(len, fi_dom.posts))) + ' characters')
print('Te dom count: ' + str(te_dom.posts.count()) + ' - ' + str(sum(map(len, te_dom.posts))) + ' characters')
print('Ti dom count: ' + str(ti_dom.posts.count()) + ' - ' + str(sum(map(len, ti_dom.posts))) + ' characters')
```
