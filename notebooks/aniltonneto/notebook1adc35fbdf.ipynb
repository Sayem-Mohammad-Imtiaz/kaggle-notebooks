# The presentation of how the chess league does

```python
labels = ["resign","mate","outoftime","draw"]
sizes = [11147,6325,1680,906]
pcts = [f'{s} {l}\n({s*100/sum(sizes):.2f}%)' for s,l in zip(sizes, labels)]
donuts = df[['victory_status']].value_counts()


donuts.plot.pie(subplots=True,figsize=(11, 6),labels=pcts,legend=[]);
```

# Most Played moves

```python
most_played_moves=df[['moves']].groupby(by='moves').size().sort_values(ascending=False).head(13);
most_played_moves.plot(kind='bar',color='orange');
```

```python
opening_name =df[['opening_name']].groupby(by='opening_name').size().sort_values(ascending=False).head(30)
opening_name.plot(kind='bar',color='orange');
```

```python
df.groupby(by='winner').size().plot(kind='bar');
```

# Best option to win with whites

```python
whiteBestOption = df[['winner','opening_name']][df.winner =='white']
```

```python
wbop = whiteBestOption.groupby(by='opening_name').count().reset_index().sort_values(by='winner', ascending=False).head()
wbop.plot(x='opening_name',kind='bar');
```

# Best option to win with blacks

```python
BlackBestOption = df[['winner','opening_name']][df.winner =='black']
```

```python
bbop = BlackBestOption.groupby(by='opening_name').count().reset_index().sort_values(by='winner', ascending=False).head()
bbop.plot(x='opening_name',kind='bar');
```
