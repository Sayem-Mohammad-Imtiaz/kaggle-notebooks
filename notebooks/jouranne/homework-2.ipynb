Por esta razón me quedaré con la foto de 1999 y 2000. ¿Qué tanto trascendio el cambio de milenio en la ocurrencia de suicidios?

```python
df.fillna(df.mean())
df.iloc[:,~df.columns.isin(['country'])].describe()
```

```python
dh=df[((df.year == 2000) ) & (df.sex != 'Both')]
print(dh.iloc[:,~dh.columns.isin(['country'])].describe())

dg=df[((df.year == 1999) ) & (df.sex != 'Both')]
print(dg.iloc[:,~dg.columns.isin(['country'])].describe())
```

```python
dg=df[((df.year == 1999) ) & (df.sex != 'Both')]
dg.iloc[:,~dg.columns.isin(['country'])].describe()
```

Se puede ver que no existe información del Indice de desarrollo para el año 1999. Está podría ser una de la variables más importantes para reflejar el cambio. Pasaré a comparar 2000 con 1999.


```python
dg=df[((df.year == 1990)  ) & (df.sex != 'Both')]
dg.iloc[:,~dg.columns.isin(['country'])].describe()
```

```python
years=[1990,2000]
dp=df[df.year.isin(years)]
plt.figure(figsize=(20,20))
sns.barplot(x="suicides_no",y="country", hue="year", data = dp)
plt.xlabel("Country",fontsize=15)
plt.ylabel("Suicides",fontsize=15)
plt.title("1990 vs 2000",fontsize=15)
plt.show()
```

Comparando el monto de suicidios en cada pais. vemos que la mayoria empeoran. Veamos como se distribuyen estas variebles:

```python
f, axes = plt.subplots(1,3, figsize=(20, 4))
sns.distplot( dp["suicides_no"], ax=axes[0])
sns.distplot( dp["population"], ax=axes[1])
sns.distplot( dp["gdp_per_capita ($)"], ax=axes[2])
```

```python

corr=dp.iloc[:,~dp.columns.isin(['generation','country-year','year'])].corr()
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Vemos que entre los paises con mayor número de suicidios en el periodo de 1986 a 2016 se encuentran Rusia, Brasil, Estados Unidos, los cuales tienen una mayor area geográfica (posible mayor exposición por mayor población) y Japón el cual ha mantenido un indeseable renombre en este tema.
