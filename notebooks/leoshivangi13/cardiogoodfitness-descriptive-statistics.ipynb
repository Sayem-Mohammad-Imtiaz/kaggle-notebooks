```python
#Binomial Distribution: Problem statement: Bank issues credit card to customers, 
#Changes of paying bill on time was 60% and if 7 accounts is selected ar rabdom, construct
#binomial probability distribution of account paying on time.
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
n=7 #total no of trials
```

```python
p=0.6 #probability of paying on time
```

```python
k=np.arange(0,8) #Creating sample space for which we will predict probability
```

```python
k
```

```python
binomial=stats.binom.pmf(k,n,p)
```

```python
binomial
```

```python
plt.plot(k,binomial,'o-')
plt.title("binomial: n=%i, p=%.2f" %(n,p),fontsize=11)
plt.xlabel("No of success")
plt.ylabel("Probability of success")
plt.show()
```

```python
sum(binomial)
```

```python
#Poissons Distribution: 6 customers arrive in 2 miniutes: 
#(a) Probability that 4 customers arrive in 1 min
#(b) Probability that more than 3 customer arrive in 1 min
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
rate=3
```

```python
n=np.arange(0,20)  #Instead of this 20 we can use any number this is just to consider finite data points.
#If make too low then leaving some probability for more than 20 and if too high then counting too many zeros
```

```python
poisson=stats.poisson.pmf(n,rate)
```

```python
poisson
```

```python
plt.plot(n,poisson,'o-')
plt.xlabel("no of customers arriving in a minute")
plt.ylabel("probability of no of customer arriving in a minute")
plt.show()
```

```python
#Normal Distribution
#Mean weight=0.295kg of a cereal pack with standard deviation of 0.025.
```

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
#(a) less than 0.280kg
stats.norm.cdf(0.280,loc=0.295,scale=0.025)
#loc is same as mean
#scale is same as standard deviation
```

```python
#(b) more than 0.350kg
1-stats.norm.cdf(0.350,loc=0.295,scale=0.025)
```

```python
#(c) between 260 and 340
stats.norm.cdf(0.340,loc=0.295,scale=0.025)-stats.norm.cdf(0.260,loc=0.295,scale=0.025)
```
