```python
data = pd.read_csv('/kaggle/input/water-potability/water_potability.csv')
data.head()
```

```python
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(data.corr(), annot = True, linewidths=.5, fmt= '.1f' , ax=ax)
```

```python
data.plot(kind = "hist", grid = True, x = "Hardness", y = "Solids")
plt.xlabel("Solids")
plt.ylabel("Hardness")
plt.title("Hardness and Solids")
```

```python
data.Hardness.plot(kind = "line", color = "g",  linewidth=1, alpha = 0.7 , grid = True , linestyle = "-")
plt.legend(loc="upper right")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("line plot")
```
