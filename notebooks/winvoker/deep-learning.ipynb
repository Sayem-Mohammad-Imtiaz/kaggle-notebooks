```python
data = pd.read_csv("/kaggle/input/breast-cancer-wisconsin-data/data.csv")
data.head()
```

```python
data.drop(["id","Unnamed: 32"],axis=1,inplace=True)
data
```

```python
data.diagnosis = [1 if each=='M' else 0 for each in data.diagnosis] # 1 hasta 
```

```python
import matplotlib.pyplot as plt

hasta = data[data.diagnosis == 1]
normal = data[data.diagnosis == 0]

plt.figure(figsize=(9,9))
plt.scatter(hasta.radius_mean,hasta.compactness_mean,color="blue",label= "male",alpha=0.5)
plt.scatter(normal.radius_mean,normal.compactness_mean,color="green",label= "female",alpha=0.5)
plt.legend()
plt.xlabel("MeanFreq")
plt.ylabel("Meanfun")
plt.show()
```

```python
y = data.diagnosis.values
x = data.drop(["diagnosis"],axis=1)
```

```python
from sklearn.model_selection import train_test_split

x_train , x_test , y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=1)
x_train.shape
```

```python
from keras.models import Sequential 
from keras.layers import Dense

model = Sequential()
model.add(Dense(units = 128,activation = 'sigmoid', input_dim = x_train.shape[1]))
model.add(Dense(units = 128,activation = 'sigmoid', input_dim = x_train.shape[1]))
model.add(Dense(units = 1,activation = 'sigmoid'))
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
model.fit(x_train,y_train,epochs=10)
```

```python
y_predicted = model.predict(x_test)
y_predicted
```

```python
y_predicted
for i,t in enumerate(y_predicted):
    if y_predicted[i] < 0.5:
        y_predicted[i] = int(0)
    else:
        y_predicted[i] = int(1)
y_predicted
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_predicted,y_test))
```
