# Windows-Store Data : Analyzation & Visualization

This dataset is basically about the app with highest rating and price of that app. Column of dataset are:
* Name: Name of the app.
* Rating: Rating for the app.
* No of People Rated : No of people who rated the app.
* Category : Category of the app.
* Date. : Date when it is posted.
* Price. : Price of the app.

In this notebook, I visualize the dataset. There are only six column in the dataset so we don't have to do large analyzing in this dataset. Beacuse only one categorical column it have. So I visualize it by choosing different columns.

Let's start visulaizing the dataset by selecting different column in my dataset.

```python
import plotly.express as px
```

```python
fig = px.scatter(data, x="Rating", y="Price", 
                 color="Rating", 
                 size = "No of people Rated",
                 hover_data=['No of people Rated','Rating', 'Category', 'Date'], 
                 title = "Visualization of Window Store")
fig.show()
```

```python
fig = px.scatter(data, x="No of people Rated", y="Price", 
                 color="Rating", 
                 size = "No of people Rated",
                 hover_data=['Rating', 'Category', 'Date'], 
                 title = "Visualization of Window Store")
fig.show()
```

```python
fig3 = px.scatter(data, x="Date", y="Name", 
                 color="Rating", 
                 size = "No of people Rated",
                 hover_data=['Price','Rating', 'Category'], 
                 title = "Visualization of Window Store")
fig3.show()
```
