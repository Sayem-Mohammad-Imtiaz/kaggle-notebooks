Replace missing values with the mean by Imputer.

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='mean')
```

```python
data_day["PM2.5"] = imp.fit_transform(data_day[['PM2.5']])
data_day["PM10"] = imp.fit_transform(data_day[['PM10']])
data_day["NO"] = imp.fit_transform(data_day[['NO']])
data_day["NO2"] = imp.fit_transform(data_day[['NO2']])
data_day["NOx"] = imp.fit_transform(data_day[['NOx']])
data_day["NH3"] = imp.fit_transform(data_day[['NH3']])
data_day["CO"] = imp.fit_transform(data_day[['CO']])
data_day["SO2"] = imp.fit_transform(data_day[['SO2']])
data_day["O3"] = imp.fit_transform(data_day[['O3']])
data_day["Benzene"] = imp.fit_transform(data_day[['Benzene']])
data_day["Toluene"] = imp.fit_transform(data_day[['Toluene']])
data_day["Xylene"] = imp.fit_transform(data_day[['Xylene']])
data_day["AQI"] = imp.fit_transform(data_day[['AQI']])
print("-------------------Dealing with NaN values in City Day Dataset successfully completed--------------------")
data_hour["PM2.5"] = imp.fit_transform(data_hour[['PM2.5']])
data_hour["PM10"] = imp.fit_transform(data_hour[['PM10']])
data_hour["NO"] = imp.fit_transform(data_hour[['NO']])
data_hour["NO2"] = imp.fit_transform(data_hour[['NO2']])
data_hour["NOx"] = imp.fit_transform(data_hour[['NOx']])
data_hour["NH3"] = imp.fit_transform(data_hour[['NH3']])
data_hour["CO"] = imp.fit_transform(data_hour[['CO']])
data_hour["SO2"] = imp.fit_transform(data_hour[['SO2']])
data_hour["O3"] = imp.fit_transform(data_hour[['O3']])
data_hour["Benzene"] = imp.fit_transform(data_hour[['Benzene']])
data_hour["Toluene"] = imp.fit_transform(data_hour[['Toluene']])
data_hour["Xylene"] = imp.fit_transform(data_hour[['Xylene']])
data_hour["AQI"] = imp.fit_transform(data_hour[['AQI']])
print("-------------------Dealing with NaN values in City Hour Dataset successfully completed--------------------")
data_station_day["PM2.5"] = imp.fit_transform(data_station_day[['PM2.5']])
data_station_day["PM10"] = imp.fit_transform(data_station_day[['PM10']])
data_station_day["NO"] = imp.fit_transform(data_station_day[['NO']])
data_station_day["NO2"] = imp.fit_transform(data_station_day[['NO2']])
data_station_day["NOx"] = imp.fit_transform(data_station_day[['NOx']])
data_station_day["NH3"] = imp.fit_transform(data_station_day[['NH3']])
data_station_day["CO"] = imp.fit_transform(data_station_day[['CO']])
data_station_day["SO2"] = imp.fit_transform(data_station_day[['SO2']])
data_station_day["O3"] = imp.fit_transform(data_station_day[['O3']])
data_station_day["Benzene"] = imp.fit_transform(data_station_day[['Benzene']])
data_station_day["Toluene"] = imp.fit_transform(data_station_day[['Toluene']])
data_station_day["Xylene"] = imp.fit_transform(data_station_day[['Xylene']])
data_station_day["AQI"] = imp.fit_transform(data_station_day[['AQI']])
print("-------------------Dealing with NaN values in Station Day Dataset successfully completed--------------------")
data_station_hour["PM2.5"] = imp.fit_transform(data_station_hour[['PM2.5']])
data_station_hour["PM10"] = imp.fit_transform(data_station_hour[['PM10']])
data_station_hour["NO"] = imp.fit_transform(data_station_hour[['NO']])
data_station_hour["NO2"] = imp.fit_transform(data_station_hour[['NO2']])
data_station_hour["NOx"] = imp.fit_transform(data_station_hour[['NOx']])
data_station_hour["NH3"] = imp.fit_transform(data_station_hour[['NH3']])
data_station_hour["CO"] = imp.fit_transform(data_station_hour[['CO']])
data_station_hour["SO2"] = imp.fit_transform(data_station_hour[['SO2']])
data_station_hour["O3"] = imp.fit_transform(data_station_hour[['O3']])
data_station_hour["Benzene"] = imp.fit_transform(data_station_hour[['Benzene']])
data_station_hour["Toluene"] = imp.fit_transform(data_station_hour[['Toluene']])
data_station_hour["Xylene"] = imp.fit_transform(data_station_hour[['Xylene']])
data_station_hour["AQI"] = imp.fit_transform(data_station_hour[['AQI']])
print("-------------------Dealing with NaN values in Station Hour Dataset successfully completed--------------------")
```
