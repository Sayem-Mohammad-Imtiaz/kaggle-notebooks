# INDIAN SCHOOL EDUCATION STATISTICS | Analysis and Visualization

As the name of notebook, **INDIAN SCHOOL EDUCATION STATISTICS** it clearly shows that in this notebook we will do some analysization and visualization part. By looking at the csv of all the dataset there are seven csv files that have data in categories i.e dropout, schools with facility, boys toilet and many more. So let's import these datasets on by one and looking into thses so that we will come to know which dataset have categorical or have only integers value. Because we have to do some analyzation first then we will do some visualization part of this dataset. First we have to import some **packages** :

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

**Importing Datasets :**

```python
data_dropout = pd.read_csv('../input/indian-school-education-statistics/dropout-ratio-2012-2015.csv')
data_gross = pd.read_csv('../input/indian-school-education-statistics/gross-enrollment-ratio-2013-2016.csv')
data_comps = pd.read_csv('../input/indian-school-education-statistics/percentage-of-schools-with-comps-2013-2016.csv')
data_electricity = pd.read_csv('../input/indian-school-education-statistics/percentage-of-schools-with-electricity-2013-2016.csv')
data_water_facility = pd.read_csv('../input/indian-school-education-statistics/percentage-of-schools-with-water-facility-2013-2016.csv')
data_boys_toilet = pd.read_csv('../input/indian-school-education-statistics/schools-with-boys-toilet-2013-2016.csv')
data_girls_toilet = pd.read_csv('../input/indian-school-education-statistics/schools-with-girls-toilet-2013-2016.csv')
```

**Glimpse of Datasets :**

```python
print("________________________________________Starting of Datasets_________________________________")
print()
print("\t\t\t_________________Dropout Ratio Dataset______________________")
print()
print(data_dropout.head())
print()
print("\t\t\t________________Gross Enrollment Dataset_______________________")
print()
print(data_gross.head())
print()
print("\t\t\t________________Percentage of Schools with Comps Dataset____________________")
print()
print(data_comps.head())
print()
print("\t\t\t___________________Percentage of Schools with Electricity_______________________")
print()
print(data_electricity.head())
print()
print("\t\t\t____________________Percentage of Schools with Water Facility______________________")
print()
print(data_water_facility.head())
print()
print("\t\t\t____________________Schools with Boys Toilets______________________")
print()
print(data_boys_toilet.head())
print()
print("\t\t\t__________________Schools with Girls Toilets_________________________")
print()
print(data_girls_toilet.head())
print("_______________________________________Ending Of Datasets_______________________________________")
```

Guys we are going to make some visualizations on each dataset we have. 
* Histogram
* Heatmap

```python
data_electricity.hist(figsize = (10, 10))
plt.show()
```

Here is the heatmap of Dataset :

```python
plt.figure(figsize = (14, 14))
sns.heatmap(data_electricity.corr(), annot = True, cmap = 'coolwarm')
plt.show()
```

```python
data_comps.hist(figsize = (10, 10))
plt.show()
```

```python
plt.figure(figsize = (14, 14))
sns.heatmap(data_comps.corr(), annot= True)
plt.show()
```

```python
data_water_facility.hist(figsize = (10, 10))
plt.show()
```

```python
plt.figure(figsize = (14, 14))
sns.heatmap(data_water_facility.corr(), annot= True, cmap = 'coolwarm', center = 0)
plt.show()
```

```python
data_boys_toilet.hist(figsize = (10, 10))
plt.show()
```

```python
plt.figure(figsize = (14, 14))
sns.heatmap(data_boys_toilet.corr(), annot= True, center = 0)
plt.show()
```

```python
data_girls_toilet.hist(figsize = (10, 10))
plt.show()
```

```python
plt.figure(figsize = (14, 14))
sns.heatmap(data_girls_toilet.corr(), annot= True, cmap = 'coolwarm')
plt.show()
```
