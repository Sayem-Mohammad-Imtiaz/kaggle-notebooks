First we have to know about the User information i.e User Name, User Information, User is verified or not etc.


**User is Verified or not.** Let's look at this column by graph plotting.

As we all know it has only two categories i.e True and False.

```python
data["user_verified"].unique()
```

Let's look at the number of user verified and number of user not verified.

```python
is_verify = len(data[data["user_verified"]==True])
is_not_verify = len(data[data["user_verified"]==False])
```

```python
print("Number of User Verified", is_verify)
print("Number of User Not Verified", is_not_verify)
```

Now look at the percentage of verfied user and not verified user.

```python
print("Percentage of User Verified: {:.2f}%".format((is_verify / (len(data.user_verified))*100)))
print("Percentage of User Not verified : {:.2f}%".format((is_not_verify / (len(data.user_verified))*100)))
```

So, by this there are only 12% users are verified. Rest of is not verified.

Let's look at the plot of user verified.

```python
sns.countplot(data.user_verified)
plt.show()
```
