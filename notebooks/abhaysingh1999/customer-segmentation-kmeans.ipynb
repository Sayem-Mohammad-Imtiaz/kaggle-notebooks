Checking for Outliers 

```python
sns.boxplot(y = "Annual Income (k$)",data = data)
```

```python
sns.distplot(data["Annual Income (k$)"])
```

```python
Q1 = data['Annual Income (k$)'].quantile(0.25)
Q3 = data['Annual Income (k$)'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# As we know that we can take value upyo range of (Q1-1.5*QPR, Q3+1.5*QPR)
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
print('lower limit is', lower_limit)
print('upper limit is', upper_limit)
```

```python
outliers_detected = []
for x in data['Annual Income (k$)']:
    if ((x> upper_limit) or (x<lower_limit)):
         outliers_detected.append(x)
print(' outlier in the dataset is', outliers_detected)
```

```python
# replacing it with the median value 
data.loc[data['Annual Income (k$)'] >= 137, 'Annual Income (k$)'] = data["Annual Income (k$)"].median()
```

```python
data_for_clustering = data.iloc[:,[3,4]].values
```

# For finding number of Clusters we Will Use Elbow Method

For finding number of Clusters we Will Use Elbow Method

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(i)
    kmeans.fit(data_for_clustering)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.xlabel("Number of cluster")
plt.ylabel("WCSS")
```

So it looks like at 5 there is break so number of cluster can be five so we will plot the graph using number of cluster as 5

```python
kmeans = KMeans(5)
```

```python
y = kmeans.fit_predict(data_for_clustering)
```

```python
plt.scatter(data["Annual Income (k$)"],y = data["Spending Score (1-100)"],c = y,cmap = "rainbow")
```

# As you can see all five cluster are differentiated properly so we can say that choosing number of clusters as 5 is the correct choice and hence we can divide the given data into five clusters as shown above
