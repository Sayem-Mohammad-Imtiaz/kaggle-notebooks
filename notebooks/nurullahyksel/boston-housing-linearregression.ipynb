```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set
data =pd.read_csv("../input/boston-housing-dataset/HousingData.csv")
```

```python
data = data.dropna()
data
```

```python
X =data.drop("MEDV", axis=1).values  #features
Y= data["MEDV"].values               #target

X
```

```python
# We examine some features separately from the table.
# RM-MEDV

Room_number = X[:,5]                              # we just get the RM features from the dataset
Room_number = Room_number.reshape(-1,1)           # split array into individual data
Y=Y.reshape(-1,1)
```

```python
#Charting

plt.scatter(Room_number,Y, )

plt.xlabel("Average Room number")
plt.ylabel("Average price (x1000 $)")
plt.title("The relationship between the number of rooms and the price of the house")
plt.show()
```

```python
#Regression model

from sklearn.linear_model import LinearRegression

regression = LinearRegression()

regression.fit(Room_number,Y) #fitting

regression_line = np.linspace(min(Room_number), max(Room_number)) 

plt.scatter(Room_number,Y, color="green")

plt.xlabel("Average Room number")
plt.ylabel("Average price (x1000 $)")
plt.title("The relationship between the number of rooms and the price of the house")

plt.plot(regression_line, regression.predict(regression_line), color="black", linewidth = 3)
plt.show()
```

```python
#Linear Regression

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X_train, X_test, Y_train, Y_test =train_test_split (X,Y, test_size=0.2, random_state=30)

regression_all = LinearRegression()
regression_all.fit(X_train, Y_train)
Y_prediction = regression_all.predict(X_test)
print ("R^2: {}" .format(regression_all.score(X_test, Y_test)))
```

```python
#mean_squared_error

from sklearn.metrics import mean_squared_error

Y_prediction = regression_all.predict(X_test)
error = np.sqrt(mean_squared_error(Y_test, Y_prediction))

print ("error: {}".format(error))
```

```python
# Ridge

from sklearn.linear_model import Ridge

ridge = Ridge(alpha=0.1, normalize= True)
ridge.fit(X_train, Y_train)
ridge_prediction = ridge.predict(X_test)

ridge.score(X_test, Y_test)

```

```python
Alpha = np.arange(0.1,1,0.1)

regr= np.empty(len(Alpha))

for i,k in enumerate (Alpha):
    ridge1= Ridge(alpha=k, normalize= True)
    ridge1.fit(X_train, Y_train)
    ridge1_prediction= ridge1.predict(X_test)
    regr[i]= ridge1.score(X_test, Y_test)
```

```python
plt.plot(Alpha,regr, label="Ridge")
plt.title("Ridge chart")
plt.xlabel("Ridge value")
plt.ylabel("Regression")
```

```python
#Lasso

from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.1, normalize= True)
lasso.fit(X_train, Y_train)
lasso_prediction = lasso.predict(X_test)

lasso.score(X_test, Y_test)
```

```python
#Which attribute is more effective on the target?
features = data.drop("MEDV", axis=1).columns

lasso = Lasso(alpha=0.1)
lasso_co = lasso.fit(X,Y).coef_

plt.plot(range(len(features)), lasso_co)
plt.xticks(range(len(features)), features, rotation=45)
plt.xlabel("Features")
plt.ylabel("Lasso")
plt.title("Which attribute is more effective on the target?")
plt.show()
```
