```python
import pandas as pd
import numpy as np 
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv("/kaggle/input/pimaindiansdiabetescsv/pima-indians-diabetes.csv")
```

```python
df.head()
```

```python
df.columns = ["pregnant","Plasma glucose","Diastolic blood pressure",
              "Triceps skin fold thickness",
              "2-Hour serum insulin",
              "Body mass index","Diabetes pedigree function",
              "Age","Class"]
```

```python
df.head()
```

```python
x = df.drop("Class",axis=1)
y = df.Class
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
model = keras.Sequential()
model.add(Dense(12,input_dim=8,activation="relu"))
model.add(Dense(8,activation="relu"))
model.add(Dense(1,activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy",
             optimizer="adam",
             metrics =["accuracy"])
```

```python
model.fit(x,y,epochs=300,batch_size=10)
```

```python
accuracy = model.evaluate(x,y)
```

```python
predictions = model.predict_classes(x)
```
