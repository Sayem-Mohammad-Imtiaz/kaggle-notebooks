```python
X=df[['sex', 'smoking','anaemia', 'diabetes', 'high_blood_pressure', 'age', 'ejection_fraction', 'time',
     'platelets', 'creatinine_phosphokinase', 
      'serum_creatinine', 'serum_sodium']].values
y=df[['DEATH_EVENT']].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, np.array(y).ravel(), random_state=1,test_size=0.2)
```

```python
#Random Forest with train test split with 20% testing size
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import precision_score, recall_score
from sklearn.metrics import f1_score

List_estim=[10,100,200]
List_depth=[2,10,15]
List_leaf_nodes=[10,20,40]
for a in List_estim:
    for b in List_depth:
        for c in List_leaf_nodes:
            clf = RandomForestClassifier(n_estimators = a, max_depth = b, max_leaf_nodes = c).fit(X_train, y_train)

            pred = clf.predict(X_test)
            
            print("Model parameters n_estimators:%0.0f, max_depth:%0.0f, max_leaf_nodes:%0.0f," % (a,b,c))
            print('Accuracy Score : {:.2f}%'.format(clf.score(X_test, y_test)*100))
            print('\nf1 Score : ', f1_score(y_test, pred))
            print('\nPrecision Score : ', precision_score(y_test, pred))
            print('\nRecall Score : ', recall_score(y_test, pred))
            print('-------------------------------------' )
```

```python
#Random Forest with Cross fold validation
List_estim=[10,100,200]
List_depth=[2,10,15]
List_leaf_nodes=[10,20,40]
for a in List_estim:
    for b in List_depth:
        for c in List_leaf_nodes:
            clf=RandomForestClassifier(n_estimators = a, max_depth = b, max_leaf_nodes = c)
            scores = cross_val_score(clf, X, np.array(y).ravel(), cv=5)
            print("Model parameters n_estimators:%0.0f, max_depth:%0.0f, max_leaf_nodes:%0.0f," % (a,b,c))
            print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
            print('-------------------------------------' )
```
