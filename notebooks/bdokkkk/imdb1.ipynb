```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import re

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv')
```

```python
df.head()
```

```python
rm_html = lambda x: re.sub(r'(<.*>)','',x)
rm_dble_slash = lambda x: re.sub(r'(\\)',' ',x)
df['review'] = df['review'].apply(rm_html)
df['review'] = df['review'].apply(rm_dble_slash)
df['sentiment'] = df['sentiment'].replace({ 'negative': 0, 'positive': 1})
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer


vectorizer = CountVectorizer(min_df=0, lowercase=True,stop_words='english',max_features=1500)
vectorizer.fit(df['review'])
```

```python
X = vectorizer.transform(df['review']).toarray()
Y = df['sentiment'].to_numpy()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

classifier = LogisticRegression()
classifier.fit(X_train, y_train)
score = classifier.score(X_test, y_test)

print("Accuracy:", score)
```

```python
from sklearn import metrics

table = metrics.confusion_matrix(classifier.predict(X_test),y_test, labels=[0,1])

pd.DataFrame(table)
```
