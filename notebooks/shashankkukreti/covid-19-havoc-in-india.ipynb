```python
data.corr()
```

* now we can see that :-
* with increase in **total cases** **discharged** cases also **increases**.
* **death count** also **increase** with **total cases**

```python
plt.figure(figsize=(15,10))
px.bar(data,'State/UTs','Total Cases',color = 'Death Ratio (%)')
```

* **Maharashtra** despite have the most cases nearly 6.2M managed to keep their **death ratio below 2%**
* whereas **kerala** even managed it more better as even having the second most number of total cases they managed the **death ratio below 1%**
* if we see the **punjab** death ratio it is the highest of **2.5%** even having **0.5m cases**
* This shows that punjab need more better facilities


```python
px.bar(data,'State/UTs','Active',color = 'Active Ratio (%)')
```

* Although the active case is most in **maharashtra** but the active cases % is most in **northeastern region**.
* this shows that Covid-19 is rampant there and we need to move our resources to that region also.

```python
px.scatter(data,'Active Ratio (%)','Death Ratio (%)',color = 'State/UTs')
```

* **Punjab** has the highest death ratio
* **Mizoram** has the highest active ratio
* **daman and diu** has the lowest active and death ratio

```python
px.scatter(data,'Total Cases','Active',size='Active Ratio (%)',color="Active Ratio (%)",hover_name ='State/UTs',log_x = True,size_max = 60)
```

* Again this shows that northeastern region having highest active % while cases are not very high
* while maharastra have high active cases but their active % is low

* **Basic points to consider:-
* States with good healthcare facilities like maharashtra and kerala managed the pandemic well in their respective states
* Whereas we need to work on healthcare facilities in states like punjab
* And In north_eastern region Covid_19 is going rampant now. so we also have to focus there as they earlier didn't need to managed it now they have to fight it**
