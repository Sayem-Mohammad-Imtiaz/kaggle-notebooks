```python
import pandas as pd
import re


'Values for TASK 2'
total1 = wrong1 = correct1 = 0
wrongdata1 = pd.DataFrame()
wrongdata2 = pd.DataFrame()
total3 = wrong3 = correct3 = 0
wrongdata3 = pd.DataFrame()
total4 = wrong4 = correct4 = 0
wrongdata4 = pd.DataFrame()
wrongdata5 = pd.DataFrame()
total6 = wrong6 = reg6 =unreg6= 0
wrongdata6 = pd.DataFrame()


'TASK 2 Functions'

def businessrule1():
    'Businessrule 1: "Vehicle Make" column should not be "TOYOT"'
    global total1,wrong1,correct1,wrongdata1
    for line in dataclear['Vehicle Make']:
        total1 += 1
        if line != 'TOYOT':
            correct1 += 1
        else:
            wrong1 += 1   
    wrongdata1=pd.concat([wrongdata1,dataclear.loc[(dataclear['Vehicle Make']=='TOYOT')]]) 
    
def businessrule2(): 
    'Businessrule 2: No fully nulls columns.'
    global wrongdata2
    wrongdata2=pd.concat([wrongdata2,dataclear])
    
def businessrule2_empt_col():
    'Function working only after func "businessrule2" and finding empty columns'
    emptycollist=[]
    for col in list(wrongdata2):
        if wrongdata2[col].value_counts().empty:
            emptycollist.append(col)
        else:
            pass
    print(wrongdata2[emptycollist])    
    
def businessrule3():
    'Businessrule 3: "Registration state": should be only from ANSI standard INCITS 38:2009'
    global total3,wrong3,correct3,wrongdata3
    wronglist=[]
    statecodes=['AL','AK','AZ','AR','CA','CO','CT','DE','DC','FL','GA','HI','ID','IL','IN',
            'IA','KS','KY','LA','ME','MD','MA','MI','MN','MS','MO','MT','NE','NV','NH',
            'NJ','NM','NY','NC','ND','OH','OK','OR','PA','RI','SC','SD','TN','TX','UT',
            'VT','VA','WA','WV','WI','WY','AS','GU','MP','PR','VI','UM','FM','MH','PW']
    for line in dataclear['Registration State']:
        total3 += 1
        if line in statecodes:
            correct3 += 1
        else:
            wrong3 += 1
            wronglist.append(line)
    wronglist=list(set(wronglist))
    for code in wronglist:
        wrongdata3=pd.concat([wrongdata3,dataclear.loc[(dataclear['Registration State']==code)]]) 
    
def businessrule4():
    'Businessrule 4: "Plate type" only PAS, COM.'
    global total4,wrong4,correct4,wrongdata4
    wronglist=[]
    for line in dataclear['Plate Type']:
        total4 += 1
        if line=='PAS' or line=='COM':
            correct4 += 1
        else:
            wrong4 += 1
            wronglist.append(line)
    wronglist=list(set(wronglist))
    for plate in wronglist:
        wrongdata4=pd.concat([wrongdata4,dataclear.loc[(dataclear['Plate Type'] ==plate)]])
         
def businessrule5():
    'Businessrule 5:  Finding min/max dates'
    global wrongdata5
    wrongdata5=pd.concat([wrongdata5,dataclear])

def businessrule6():
    'Businessrule 6: No unregistered vehicle in the "Unregistered Vehicle" column'
    global total6,wrong6,reg6,unreg6,wrongdata6
    for line in dataclear['Unregistered Vehicle?']:
        total6 += 1
        if line==0:
            reg6 += 1
        elif line==1:
            unreg6+=1
        else:
            wrong6 += 1
    wrongdata6=pd.concat([wrongdata6,dataclear.loc[(dataclear['Unregistered Vehicle?'] !=0)]]) 
    
    
'Main part: reading csv file with chunk'

url='/kaggle/input/nyc-parking-tickets/Parking_Violations_Issued_-_Fiscal_Year_2017.csv'  
dataclear = pd.DataFrame() #temporary dataframe for saving clear data
for data in pd.read_csv(url, chunksize=5000000): 
    data['Issue Date']= pd.to_datetime(data['Issue Date'])
    dataclear=data[(data['Issue Date']>='2017-01-01')&(data['Issue Date']<='2017-12-31')]
    'Functions executing Task 2'
    businessrule1()
    businessrule2()
    businessrule3()
    businessrule4()
    businessrule5()
    businessrule6()
    

'OUTPUT TASK 2'
'Output Businessrule 1'
print('\n\n\nTASK 2\n\n')
print('Businessrule 1: "Vehicle Make" column should not be "TOYOT"')
print('total rows:', total1, 'correct rows:', correct1, 'wrong rows:', wrong1)
print('Table with wrong data')
print(wrongdata1[['Summons Number','Plate ID','Issue Date','Vehicle Make']])

'Output Businessrule 2'
print('\n\nBusinessrule 2: No fully nulls columns')
print('Table with emty columns:')
businessrule2_empt_col()

'Output Businessrule 3'
print('\n\nBusinessrule 3: "Registration state": should be only from ANSI standard INCITS 38:2009')
print('total rows:', total3, 'correct rows:', correct3, 'wrong rows:', wrong3)
print('Table with wrong data')
print(wrongdata3[['Summons Number','Plate ID','Issue Date','Registration State']])

'Output Businessrule 4'
print('\n\nBusinessrule 4: "Plate type" only PAS, COM.')
print('total rows:', total4, 'correct rows:', correct4, 'wrong rows:', wrong4)
print('Table with wrong data')
print(wrongdata4[['Summons Number','Plate ID','Issue Date','Plate Type']])

'Output Businessrule 5'
print('\n\nBusinessrule 5: Finding min/max dates')
print('Minimum Issue date:',min(wrongdata5['Issue Date'].dt.date))
print('Maximum Issue date:',max(wrongdata5['Issue Date'].dt.date))

'Output Businessrule 6'
print('\n\nBusinessrule 6: No unregistered vehicle in the "Unregistered Vehicle" column')
print('total rows:', total6)
print('registered vehicle:', reg6)
print('unregistered vehicle:',unreg6)
print('unidentified vehicle:',wrong6)
print('Table with wrong data')
print(wrongdata6[['Summons Number','Plate ID','Issue Date','Unregistered Vehicle?']])
```
