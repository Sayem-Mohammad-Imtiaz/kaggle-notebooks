```python
import pandas as pd
df = pd.read_csv("../input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv")
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df['gender'] = labelencoder.fit_transform(df['gender'])
df['ever_married'] = labelencoder.fit_transform(df['ever_married'])
df['work_type'] = labelencoder.fit_transform(df['work_type'])
df['Residence_type'] = labelencoder.fit_transform(df['Residence_type'])
df['smoking_status'] = labelencoder.fit_transform(df['smoking_status'])

df = df.drop(columns = ["id"])
df.head()
```

```python
df = df.dropna()
```

```python
df.isnull().values.any()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler 
import numpy as np

df["age"] = MinMaxScaler().fit_transform(np.array(df["age"]).reshape(-1,1)) 
df["avg_glucose_level"] = MinMaxScaler().fit_transform(np.array(df["avg_glucose_level"]).reshape(-1,1)) 
df["bmi"] = MinMaxScaler().fit_transform(np.array(df["bmi"]).reshape(-1,1)) 
```

```python
df.head()
```

```python
X = df.iloc[:, :10]
y = df.iloc[:, 10:]
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 69)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import Activation, Dense
from keras.models import Sequential
```

```python
X.head()
```

```python
#define model
model = Sequential()
```

```python
#define firsts hidden and visible layer
model.add(Dense(50, input_dim = 10, activation = "relu", kernel_initializer  = "he_uniform" ))
```

```python
# define second hidden layer
model.add(Dense(25, activation = "relu"))
```

```python
#define output layer
model.add(Dense(1, activation = "sigmoid"))
```

```python
#define loss and optimizer
model.compile(loss = "binary_crossentropy", optimizer = "adam")
```

```python
model.fit(X_train, y_train, epochs = 10)
```

```python
from sklearn.metrics import roc_auc_score
y_pred1=model.predict(X_test)

print(roc_auc_score(y_test,y_pred1))
```
