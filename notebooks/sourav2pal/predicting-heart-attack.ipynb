# Feature Engineering

In this section we will be trying to determine features that are relevent with respect to the problem at hand. This includes exploring the different types of feature-selection techniques.

```python
continuous = heart[continuous_features]
categorical = heart[categorical_features]
```

```python
from sklearn.feature_selection import SelectKBest, f_classif, chi2, mutual_info_classif
from scipy.stats import kendalltau

annova_score_model = SelectKBest(score_func=f_classif, k="all").fit(continuous.values, heart.output.values)
annova_scores=pd.DataFrame(annova_score_model.scores_, index=continuous.columns, columns=['F Scores'])
annova_scores = annova_scores.sort_values(by='F Scores', ascending=False).reset_index()

chi2_score_model = SelectKBest(score_func=chi2, k="all").fit(categorical.values, heart.output.values)
chi2_scores=pd.DataFrame(chi2_score_model.scores_, index=categorical.columns, columns=['Chi2'])
chi2_scores = chi2_scores.sort_values(by='Chi2', ascending=False).reset_index()

mut_clas_score_model=SelectKBest(score_func=mutual_info_classif, k="all").fit(categorical.values, heart.output.values)
mut_clas_scores=pd.DataFrame(mut_clas_score_model.scores_, index=categorical.columns, columns=['Mutual Information'])
mut_clas_scores=mut_clas_scores.sort_values(by="Mutual Information", ascending=False).reset_index()

tau_scores_list = []
for feat in continuous.columns:
    score, _=kendalltau(continuous[feat], heart.output)
    tau_scores_list.append(abs(score))
tau_scores = (pd.DataFrame({'index':continuous.columns, 'Tau':tau_scores_list})
                .sort_values(by="Tau", ascending=False).reset_index(drop=True))


fig_score, ax = plt.subplots(2, 2, figsize = (20, 14))
sns.barplot(x='index', y='F Scores', data=annova_scores, ax=ax[0][0])
sns.barplot(x='index', y='Chi2', data=chi2_scores, ax=ax[0][1])
sns.barplot(x='index', y='Tau', data=tau_scores, ax=ax[1][0])
sns.barplot(x='index', y='Mutual Information', data=mut_clas_scores, ax=ax[1][1])
```

### The Above graphs gives an idea regarding the relative importance of the different continuous and categorical features are present in the data with respect to the categorical target.
