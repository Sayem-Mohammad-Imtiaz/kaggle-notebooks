```python
import seaborn as sns
import numpy as np 
import pandas as pd
```

```python
data = pd.read_csv('/kaggle/input/top50spotify2019/top50.csv',encoding='latin-1')
```

```python
data.head()
```

Popularity of Songs will give us the desired properties for songs

```python
first_100 = data.sort_values('Popularity',ascending = False).head(100)
```

```python
first_100
```

```python
ax = sns.barplot(x="Artist.Name", y="Beats.Per.Minute", data=first_100)
```

100-125 beats per minute seems ideal for songs.

We can look out the changes of differences of values by looking standard deviation. Which are more likely changing to ratio it by 

```python
x = first_100.groupby('Genre').mean()
y = first_100.groupby('Genre').std()
```

```python
(y / x).sort_values('Popularity')
```

We see that most consistent Genres by means of popularity are "brostep", "latin", "dfw rap". This table shows us the constancy of genre.
