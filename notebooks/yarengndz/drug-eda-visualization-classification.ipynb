dataframe has some outlier values in Na_to_K column.These can be organic or inorganic values,these are so critical to make operations.Let's stay steady for this step

```python
def detect_outlier(data):
    q1 = np.quantile(data, 0.25)
    q3 = np.quantile(data, 0.75)
    IQR = q3 - q1
    low = q1 - (1.5 * IQR)
    high = q3 + (1.5 * IQR)
    return high, low
```

```python
df[["Na_to_K"]].describe().T
```

Minimum and maximum bound of this column was 6.0 and 38.0.After the outlier detection,the boundary is expanded from  -3.5 to 32.5.Research the outlier values

```python
high, low = detect_outlier(df["Na_to_K"])
print(high, "--", low)
```

```python
df[df["Na_to_K"] < low]
```

```python
df[df["Na_to_K"] > high]
```

Leave them same,the outlier is seen as organic event

# 6. Data Classification by DecisionTreeClassifier

First of all, we should look the dataframe and the categorical values to make them digitalize.

```python
df.head()
```

```python
for i in df.Sex.values:
    if i  == 'M':
        df.Sex.replace(i, 0, inplace=True)
    else:
        df.Sex.replace(i, 1, inplace=True)
for i in df.BP.values:
    if i == 'LOW':
        df.BP.replace(i, 0, inplace=True)
    elif i == 'NORMAL':
        df.BP.replace(i, 1, inplace=True)
    elif i == 'HIGH':
        df.BP.replace(i, 2, inplace=True)
for i in df.Cholesterol.values:
    if i == 'LOW':
        df.Cholesterol.replace(i, 0, inplace=True)
    else:
        df.Cholesterol.replace(i, 1, inplace=True)
```

Let's control the dataframe after these for's to make all df numeric

```python
df.head()
```

Let's create two df one is for properties, the other one is for target properties

```python
X_var = df[['Sex', 'BP', 'Age', 'Cholesterol', 'Na_to_K']].values # independent variable
y_var = df['Drug'].values # dependent variable
```

```python
ss = StandardScaler()
```

```python
ss.fit(X_var)
```

```python
X = ss.transform(X_var)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_var, test_size=0.3, random_state=0)
```

```python
dtree = tree.DecisionTreeClassifier(criterion="entropy", random_state=0)
dtree.fit(X_train, y_train)
```

```python
print("Train - Accuracy: ", metrics.accuracy_score(
    y_train, dtree.predict(X_train)))
print("Train - Confusion Matrix: ", metrics.confusion_matrix(
    y_train, dtree.predict(X_train)))
print("Train - Classification Report: ", metrics.classification_report(
    y_train, dtree.predict(X_train)))
print("\n****************************\n")
print("Test - Accuracy: ", metrics.accuracy_score(
    y_test, dtree.predict(X_test)))
print("Test - Confusion Matrix: ", metrics.confusion_matrix(
    y_test, dtree.predict(X_test)))
print("Test - Classification Report: ", metrics.classification_report(
    y_test, dtree.predict(X_test)))
```
