As we can see from the data set that there are different reason for 911 calls. for example "EMS: BACK PAINS/INJURY". We need categorize these reasons.


We will create a new colum with name of **Reasons**

```python
df['Reasons']=df['title'].apply(lambda x: x.split(':')[0])
```

```python
df.head()
```

Now you can see that there is another column **Reasons** , we can use this column to further present the data on the basis of reasons for 911 call.

Total number of calls for EMS, Traffic and Fire are as below

```python
df['Reasons'].value_counts()
```

Count plot for total calls on the basis of the **Reasons**

```python
sns.countplot(x='Reasons', data=df)
```

Now we need to present the data on the basis of date.

First we need to check the data type of date in data set

```python
type(df['timeStamp'].iloc[0])
```

As we can see that the data type is str. We need to convert this data type to time so we can futher break it down to presdent the data on the basis of Day, Month, Year.

```python
df['timeStamp']=pd.to_datetime(df['timeStamp'])
type(df['timeStamp'].iloc[0])
```

We will make separate columns for Day, Month and Year

```python
df['Day']=df['timeStamp'].apply(lambda x: x.dayofweek)
df['Day']
```

Now we need to change the day number to actual name of the day, we will do this using the map(). First we need to create a key.

```python
dmap= {0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
```

```python
df['Day']=df['Day'].map(dmap)
```

Now we will see the data set to check if the number of days has been changed to name of days.

```python
df.head()
```

The last column is now showing the name of days.

Now we will create coulmns for month,year and time


```python
df['Month']=df['timeStamp'].apply(lambda x: x.month)
df['Year']=df['timeStamp'].apply(lambda x: x.year)
df['Time']=df['timeStamp'].apply(lambda x: x.time())
```

```python
df.head()
```

Now we will present the data on the basis of Month and year

```python
sns.countplot(x='Month',data=df)
```

Total number of calls per month on the basis of reasons

```python
sns.countplot(x='Month',data=df, hue='Reasons')
```

Total number of calls per day on the basis of reasons

```python
sns.countplot(x='Day',data=df, hue='Reasons')
```
