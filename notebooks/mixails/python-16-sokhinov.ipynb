```python
df = pd.read_csv('/kaggle/input/weather-dataset/weatherHistory.csv')
df
```

```python
import statistics as stat

median_Temp=np.median(df['Temperature (C)'])
print("Медиана температуры: ", median_Temp)

median_Humidity=np.median(df['Humidity'])
print("Медиана влажности: ", median_Humidity)

median_Wind=np.median(df['Wind Speed (km/h)'])
print("Медиана скорости ветра: ", median_Wind)

mode_Temp=stat.mode(df['Temperature (C)'])
print("Мода температуры: ", mode_Temp)

mode_Humidity=stat.mode(df['Humidity'])
print("Мода влажности: ", mode_Humidity)

mode_Wind=stat.mode(df['Wind Speed (km/h)'])
print("Мода скорости ветра: ", mode_Wind)

mean_Temp=np.mean(df['Temperature (C)'])
print("Средняя температуры: ", mean_Temp)

mean_Humidity=np.mean(df['Humidity'])
print("Средняя влажности: ", mean_Humidity)

mean_Wind=np.mean(df['Wind Speed (km/h)'])
print("Средняя скорости ветра: ", mean_Wind)

max_Temp=np.max(df['Temperature (C)'])
print("Максимум температуры: ", max_Temp)

max_Humidity=np.max(df['Humidity'])
print("Максимум влажности: ", max_Humidity)

max_Wind=np.max(df['Wind Speed (km/h)'])
print("Максимум скорости ветра: ", max_Wind)

min_Temp=np.min(df['Temperature (C)'])
print("Минимум температуры: ", min_Temp)

min_Humidity=np.min(df['Humidity'])
print("Минимум влажности: ", min_Humidity)

min_Wind=np.min(df['Wind Speed (km/h)'])
print("Минимум скорости ветра: ", min_Wind)
```

Вытаскиваю из таблицы данных один день - 28.02.2006

```python
day28 = df[df['Formatted Date'].str.match('^2006-02-28')].hist
day28

```

Чисто по этому дню, так как там идут измерения температуры по часам, строю график её изменения.

```python
df_x = day28['Formatted Date'].values
df_y = day28['Temperature (C)'].values

fig, ax = plt.subplots()

ax.bar(df_x, df_y)

ax.set_xticklabels(df_x, rotation = 90)
fig.set_facecolor('floralwhite')
fig.set_figwidth(10) # ширина Figure
fig.set_figheight(6) # высота Figure
plt.show()
```
