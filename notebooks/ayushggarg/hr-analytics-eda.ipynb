## **Load the training data**

```python
df_train = pd.read_csv("/kaggle/input/hr-analytics-job-change-of-data-scientists/aug_train.csv")
print (df_train.shape)
df_train.head()
```

## **Missing values in each column**

```python
sns.set(style="darkgrid")
pd.DataFrame(df_train.isna().sum().sort_values(ascending=True), columns=['Missing Values']
            ).plot(kind='barh', figsize=(8,6))
plt.show()
```

## **Unique values in each column**

```python
sns.set(style="darkgrid")
pd.DataFrame(df_train.drop(columns=['enrollee_id']).nunique().sort_values(ascending=True), 
             columns=['Unique Values']).plot(kind='barh', figsize=(8,6), logx=True)
plt.show()
```

# **EDA of Training Data**

## **Number of employees in different categories across both the type of target**

```python
list_ = ['gender', 'relevent_experience', 'enrolled_university', 'education_level', 'major_discipline',
         'company_size', 'company_type', 'last_new_job']
sns.set(style="darkgrid")
fig, ax = plt.subplots(4, 2, figsize=(20,30))
for i, ax in zip(range(8), ax.flat):
    col = list_[i]
    sns.countplot(x=col, hue='target', data=df_train, ax=ax)
plt.show()
```

## **Heatmap (correlation) among various numerical features**

```python
plt.figure(figsize=(8,8))
sns.heatmap(df_train.drop(columns=['enrollee_id']).corr(), annot=True, square=True)
plt.show()
```

## **Number of employees with different education level across enrolled university**

```python
sns.set(style="darkgrid")
plt.figure(figsize=(8,6))
sns.countplot(x='enrolled_university', hue='education_level', data=df_train)
plt.show()
```

## **Number of employees with different company size across company types**

1. '<10' -> 0, 
1. '10/49' -> 1, 
1. '50-99' -> 2, 
1. '100-500' -> 3, 
1. '500-999' -> 4, 
1. '1000-4999' -> 5, 
1. '5000-9999' -> 6, 
1. '10000+' -> 7

```python
### drop city column, since every unique city is associated with a unique development index
df_train.drop(columns=['city'], inplace=True)
```

```python
### converting 'company_size' to ordinal feature
dict_mapping = {'<10':0, '10/49':1, '50-99':2, '100-500':3, '500-999':4, '1000-4999':5, '5000-9999':6, '10000+':7}
df_train['company_size'].replace(dict_mapping, inplace=True)

sns.set(style="darkgrid")
plt.figure(figsize=(12,8))
sns.countplot(x='company_type', hue='company_size', data=df_train)
plt.show()
```
