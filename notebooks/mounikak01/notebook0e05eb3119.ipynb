```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
df.head()
```

```python
data_numerical= df[['math score','reading score','writing score']]
```

```python
plt.figure(dpi=100)
plt.title('Correlation Analysis')
sns.heatmap(data_numerical.corr(),annot=True,lw=1)
plt.xticks(rotation=60)
plt.yticks(rotation = 60)
plt.show()
```
