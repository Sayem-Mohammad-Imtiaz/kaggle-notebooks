```python
#train_test_split
from sklearn.model_selection import train_test_split
train_X,val_X,train_y,val_y = train_test_split(X,y,test_size = 0.3,random_state = 2)
```

**Me :** By the way you will do that 100 times for find the best count of neighbor.(İt is a KNN algorithm options)

```python
#import model
from sklearn.neighbors import KNeighborsClassifier

x_list = np.arange(1,150)
score_list = []
for i in range(1,150):

    knn = KNeighborsClassifier(n_neighbors = i)

    knn.fit(train_X,train_y)
    
    score_list.append(knn.score(val_X,val_y))
```

Let's look at what is the best neighbor.

```python
plt.plot(x_list,score_list,color = "red")
```

**We are seeing '27' neighbor is the best neighbor count.**

**So,We will use our model with '27' neighbor count.**

```python
knn = KNeighborsClassifier(n_neighbors = 27)
knn.fit(train_X,train_y)
print("KNN Score : ",knn.score(val_X,val_y))
```
