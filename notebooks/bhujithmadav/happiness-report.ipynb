```python
regionWiseDF=pd.DataFrame(df.groupby('Regional indicator').count().sort_values(by='Country name',ascending=False)['Country name'])
regionWiseDF.rename(columns={'Country name':'No of countries in the region'},inplace=True)

fig=plt.figure(figsize=(10,6))
gs=fig.add_gridspec(1,1)

ax=fig.add_subplot(gs[0,0])
ax.set_facecolor('#E6E6E6')

j=0
for i in regionWiseDF['No of countries in the region']:
    ax.annotate(str(i),xy=(i,j),va='center',ha='right',fontweight='light', fontfamily='monospace',fontsize=15, color='white')
    j=j+1
    
ax.text(0,-1,'The Regions in the World and No of Countries in each Region',fontfamily='sans-serif',fontsize=20,fontweight='bold',color='#323232')

img=sns.barplot(y=regionWiseDF.index,x='No of countries in the region',data=regionWiseDF,palette='bright',ax=ax)
img.set(xlabel=None,xticklabels=[],ylabel=None)
img.tick_params(left=False,bottom=False)

for s in ['top', 'bottom', 'right']:
            ax.spines[s].set_visible(False)
```

```python
top20DF=pd.DataFrame(df.iloc[:20].groupby('Regional indicator').count().sort_values('Country name',ascending=False)['Country name'])
top20DF.rename(columns={'Country name':'No of countries in the region'},inplace=True)

bottom20DF=pd.DataFrame(df.iloc[128:].groupby('Regional indicator').count().sort_values('Country name',ascending=False)['Country name'])
bottom20DF.rename(columns={'Country name':'No of countries in the region'},inplace=True)

fig=plt.figure(figsize=(20,6))
gs=fig.add_gridspec(1,2)

fig.set_facecolor('#E6E6E6')

ax0=fig.add_subplot(gs[0,0])
ax1=fig.add_subplot(gs[0,1])

i=0
for region in top20DF.index:
    ax0.annotate(region,xy=(i,6),va='center',ha='right',fontweight='light', fontfamily='monospace',fontsize=10, color='black',rotation=90)
    i=i+1

i=0
for region in bottom20DF.index:
    ax1.annotate(region,xy=(i,6),va='center',ha='right',fontweight='light', fontfamily='monospace',fontsize=10, color='black',rotation=90)
    i=i+1

ax0.text(0,15,'Regions and the no of countries in Top 20 of the report',fontfamily='sans-serif',fontsize=13,fontweight='bold',color='#323232')
img1=sns.barplot(x=top20DF.index,y='No of countries in the region',data=top20DF,palette='bright',ax=ax0)
img1.set(xlabel=None,xticklabels=[])
img1.tick_params(left=False,bottom=False)

ax1.text(0,17,'Regions and the no of countries in Bottom 20 of the report',fontfamily='sans-serif',fontsize=13,fontweight='bold',color='#323232')
img2=sns.barplot(x=bottom20DF.index,y='No of countries in the region',data=bottom20DF,palette='pastel',ax=ax1)
img2.set(xlabel=None,xticklabels=[])
img2.tick_params(left=False,bottom=False)

ax0.text(-0.5,-1,'The Western European region has the most no of countries in Top 20',fontfamily='sans-serif',fontsize=13,fontweight='bold',color='#323232')
ax1.text(-0.5,-1,'The Sub-Saharan African region has the most no of countries in Bottom 20',fontfamily='sans-serif',fontsize=13,fontweight='bold',color='#323232')

ax0.text(0,-3,'Its interesting to note that the Sub Saharan African region has the most countries in the list but also has major no of countries in bottom half',
         fontfamily='sans-serif',fontsize=15,fontweight='light',color='#323232')

for s in ['top', 'bottom', 'right']:
    ax0.spines[s].set_visible(False)
    ax1.spines[s].set_visible(False)
```

```python
max=0
country=''
for i in df2['Country name'].unique():
    countryDF=df2[df2['Country name']==i]
    countryDF.reset_index(inplace=True)
    size=len(countryDF)
    start=countryDF.loc[0]['Life Ladder']
    end=countryDF.loc[size-1]['Life Ladder']
    cagrGrowth=(((end/start)**(1/size))-1)*100
    if cagrGrowth>max and size>7:
        max=cagrGrowth
        country=i
#print(country,' ',max)

fig=plt.figure(figsize=(20,3))
gs=fig.add_gridspec(1,1)
ax0=fig.add_subplot(gs[0,0])
ax0.set_facecolor('#F6F6F6')
ax0.set_xticklabels('')
ax0.set_yticklabels('')
ax0.tick_params(axis='both',length=0)

ax0.text(0.27,0.7,'Country with highest CAGR Growth of Happiness Score',fontfamily='monospace',fontweight='bold',fontsize=18,color='blue')
ax0.text(0.42,0.4,'Liberia - 3.67 % Growth',fontfamily='monospace',fontweight='bold',fontsize=15,color='green')
ax0.text(0.27,0.1,'*Only countries with data available for more than 7 years was considered*',fontfamily='monospace',fontweight='light',fontsize=12,color='black')

for s in ['top','bottom','left','right']:
    ax0.spines[s].set_visible(False)
```
