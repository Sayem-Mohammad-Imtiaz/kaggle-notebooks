```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import matplotlib.pyplot as plt
import pylab as pl # for showcasing the graph
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn import linear_model
from sklearn.metrics import r2_score

df = pd.read_csv('/kaggle/input/chess/games.csv')  

sub_df = df[['white_rating','black_rating']]
sub_df.head(6)

# Modelling the data 70% of the csv used for training
rand_row = np.random.rand(len(df)) < 0.5
train = sub_df[rand_row] # Training the data set
test = sub_df[~rand_row] # Testing the data set

reg = linear_model.LinearRegression()
train_x = np.asanyarray(train[['white_rating']])
train_y = np.asanyarray(train[['black_rating']])
reg.fit(train_x,train_y)

# Checking if the scatter plot is linear or not
plt.scatter(train.white_rating, train.black_rating,  color='blue')
plt.plot(train_x, reg.coef_[0][0]*train_x + reg.intercept_[0], '-r')
plt.xlabel("White")
plt.ylabel("Black")

# Testing the data 
test_x = np.asanyarray(test[['white_rating']])
test_y = np.asanyarray(test[['black_rating']])
test_y_ = reg.predict(test_x)

print("Mean Absolute Error of Model: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Accuracy of the Model: %.2f" % r2_score(test_y , test_y_))
```
