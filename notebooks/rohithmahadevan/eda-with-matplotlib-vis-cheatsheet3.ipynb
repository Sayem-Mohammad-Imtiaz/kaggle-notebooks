```python
potability_val = df.Potability.value_counts()
potable = pd.DataFrame(potability_val,index =[0,1])
potable
```

```python
plt.figure(figsize=(8,4))
plt.style.use('seaborn-notebook')

plt.bar(potable.index , potable.Potability, width = 0.1)
plt.xticks([0,1])
plt.show()
```

```python
#create a new df with potable water values


potable_water = df[df['Potability']==1]
potable_water
```

```python
agg_potable_water = potable_water.mean()
agg_potable_water
```

```python
agg_potable_water['Solids'] = agg_potable_water['Solids']/100
```

```python
agg_potable_water
```

```python
agg_potable_water = pd.DataFrame(agg_potable_water, columns =['values'])
agg_potable_water
```

```python
plt.style.use('tableau-colorblind10')
plt.figure(figsize=(10,5))

agg_potable_water.plot(kind='bar')
plt.show()
```

```python
#create a new df with potable water values


non_potable_water = df[df['Potability']==0]
non_potable_water
```

```python
agg_non_potable_water = non_potable_water.mean()
agg_non_potable_water
```

```python
agg_non_potable_water['Solids'] = agg_non_potable_water['Solids']/100
```

```python
agg_non_potable_water
```

```python
agg_non_potable_water = pd.DataFrame(agg_non_potable_water, columns =['values'])
agg_non_potable_water
```

```python
plt.style.use('bmh')
agg_non_potable_water.plot(kind='bar')
```

```python

concat_val = pd.concat([agg_potable_water,agg_non_potable_water],axis =1)
concat_val
```

```python
concat_val.columns = ['portable','non_portable']
```

```python
concat_val
```

```python
concat_val
```

```python
width = 0.35  
x = np.arange(len(concat_val.index))  

fig, ax = plt.subplots(figsize=(13,5))
rects1 = ax.bar(x - width/2, concat_val['portable'], width, label='portable')
rects2 = ax.bar(x + width/2,  concat_val['non_portable'], width, label='non_portable')


ax.set_ylabel('portability')
ax.set_title('portability of water')
ax.set_xticks(x)
ax.set_xticklabels(concat_val.index,Rotation = 30)
ax.legend()

ax.bar_label(rects1, label_type ='edge')


fig.tight_layout()

plt.show()
```
