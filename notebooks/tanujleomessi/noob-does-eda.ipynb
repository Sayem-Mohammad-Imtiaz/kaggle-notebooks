```python
#is this the part which everyone 100% knows...
import pandas as pd
import seaborn as sns
import missingno as msno
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("/kaggle/input/students-performance-in-exams/StudentsPerformance.csv")
```

# Data Exploration
* Let's dive into our data !

```python
data.head()
```

```python
data.columns
```

* Some column names are way long , I'll rename them

```python
data.rename(columns={"race/ethnicity":"race",'parental level of education':'parent_education','test preparation course':'testprep_course'},inplace=True)
```

* Let's look for missing values

```python
data.isna().sum()
```

No missing valueus !!!

```python
data.shape
```

* There are three papers(100 marks each) .
* Lets calculate their percentage and add it as a new column

```python
data['percent']=((data['math score']+data['reading score']+data['writing score'])/300)*100
data['percent']=data['percent'].round(2)
```

```python
data['parent_education'].unique()
```

* Replacing some college with bachelor's degree.
* Replacing some high school with high school.

```python
data['parent_education'].astype(str)
data['parent_education']=data['parent_education'].apply(lambda x:x.replace("some college","bachelor's degree"))
```

```python
data['parent_education']=data['parent_education'].apply(lambda x:x.replace('some high school',"high school"))
```

# Exploratory data analysis


```python
data.describe()
```

* Max percent- 100
* Min percent - 9
* The average guy scored 68.33%

```python
sns.histplot(data['percent'])
```

* Let's look at the gender distribution.

```python
data.head()
```

```python
sns.countplot(x=data['gender'],hue=data['testprep_course'])
plt.figure(figsize=(8,8))
```

* We have more females in our data.
* Females generally scored better than the males :(

```python
data.groupby('testprep_course')[['percent','math score','reading score','writing score']].agg('mean')
```

* Pretty obvious that one's who prepared for the test did well.

```python
data.groupby('gender')[['percent']].agg('mean')
```

* The females have a better performance than the male.
* Let's look at the influence of parent's education on score.

```python
plt.pie(data['parent_education'].value_counts(),labels=data['parent_education'].unique(),autopct='%0.2f%%',radius=1.2)

plt.show()
```

* As far as degrees are concerned , we can say Masters> Bachelor's> Associates > High school
*  In our data , most people have a Bachelor's degree.
* There are very few people who have studied till high school(5.9%).


```python
parent=data.groupby('parent_education')['percent'].agg('mean')
parent
```

* Clearly , parent education plays a huge role in children's future.
* Children of highly educated parents are more likely to excel.

```python
plt.figure(figsize=(8,6))
sns.barplot(x=parent.index,y=parent)
```
