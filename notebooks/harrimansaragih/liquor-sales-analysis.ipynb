```python
df = pd.read_csv('/kaggle/input/iowa-liquor-sales/Iowa_Liquor_Sales.csv')
```

```python
dfx = df.drop(columns = ['Invoice/Item Number', 'Date', 'Store Number', 'Store Name', 'Address', 'City', 'Zip Code', 'Store Location', 'County Number', 'County','Category', 'Category Name', 'Vendor Number', 'Item Number', 'Pack', 'Bottle Volume (ml)', 'State Bottle Cost', 'State Bottle Retail', 'Volume Sold (Liters)', 'Volume Sold (Gallons)'])
dfx
```

```python
dfx['Sale (Dollars)'] = dfx['Sale (Dollars)'].str.replace('$','').astype(float)
dfx
```

```python
dfx.groupby('Vendor Name')['Bottles Sold'].sum().sort_values(ascending=False).head(10).plot(kind='barh', figsize=(20,10))
plt.title('Bottles Sold')
plt.show()
```

```python
dfx.groupby('Item Description')['Bottles Sold'].sum().sort_values(ascending=False).head(10).plot(kind='barh', figsize=(20,10))
plt.title('Bottles Sold')
plt.show()
```

```python
dfx.groupby('Item Description')['Sale (Dollars)'].sum().sort_values(ascending=False).head(10).plot(kind='barh', figsize=(20,10))
plt.title('Sale (Dollars)')
plt.show()
```
