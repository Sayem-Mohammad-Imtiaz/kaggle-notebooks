```python
df_new=df.drop(columns=['POSTED_BY','LATITUDE','LONGITUDE','READY_TO_MOVE','ADDRESS','RERA','UNDER_CONSTRUCTION'])
```

```python
df_test.drop(columns=['POSTED_BY','LATITUDE','LONGITUDE','READY_TO_MOVE','ADDRESS','RERA','UNDER_CONSTRUCTION'],inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_new['BHK_OR_RK']=le.fit_transform(df_new['BHK_OR_RK'])
df_test['BHK_OR_RK']=le.fit_transform(df_test['BHK_OR_RK'])
```

```python
arr=['BHK_NO.', 'BHK_OR_RK', 'SQUARE_FT', 'RESALE']
X=df_new[arr]
y=df_new['TARGET(PRICE_IN_LACS)']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.15,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lmodel=LinearRegression()
```

```python
lmodel.fit(X_train,y_train)
```

```python
lmodel.score(X_train,y_train)
```

```python
pred=lmodel.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
from math import *

sqrt(mean_squared_error(pred,y_test))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, pred)
```
