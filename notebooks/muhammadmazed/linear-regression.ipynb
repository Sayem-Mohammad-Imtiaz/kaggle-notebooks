```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('../input/life-expectancy-who/Life Expectancy Data.csv')
df.head()
```

```python
df.info()
```

```python
df.drop(['Country', 'Status'], axis=1, inplace=True)
df.head()
```

```python
df.isnull().values.any()
```

```python
df.isna().any()
```

```python
df['Adult Mortality'].fillna(df['Adult Mortality'].median(), inplace=True)
```

```python
df['Life expectancy '].fillna(df['Life expectancy '].median(), inplace=True)
```

```python
df['Hepatitis B'].fillna(df['Hepatitis B'].median(), inplace=True)
```

```python
df['Polio'].fillna(df['Polio'].median(), inplace=True)
```

```python
plt.figure(figsize=(5,5))
plt.xlabel('Adult Mortality')
plt.ylabel('Life expectancy')
plt.scatter(df['Adult Mortality'], df['Life expectancy '])
```

```python
plt.figure(figsize=(5,5))
plt.xlabel('infant death')
plt.ylabel('Life expectancy')
plt.scatter(df['infant deaths'], df['Life expectancy '])
```

```python
plt.figure(figsize=(5,5))
plt.xlabel('under-five deaths')
plt.ylabel('Life expectancy')
plt.scatter(df['under-five deaths '], df['Life expectancy '])
```

```python
model = linear_model.LinearRegression().fit(df[['Adult Mortality', 'infant deaths', 'under-five deaths ']], df['Life expectancy '])
```

```python
model.coef_
```

```python
model.intercept_
```
