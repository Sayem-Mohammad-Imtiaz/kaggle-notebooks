```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

<span style="font-size:30px;color:#9ACD32;font-weight:bold"><center>Loading Data</center></span>

```python
df=pd.read_csv('../input/breast-cancer-wisconsin-data/data.csv')
```

```python
df
```

```python
df.info()
```

<center><span style="font-size:27px;color:#9ACD32;font-weight:bold">Checking NaNs

```python
df.isna().sum()
```

```python
df.drop(['Unnamed: 32'],axis =1,inplace = True)
```

```python
sns.countplot(df.diagnosis,palette="YlGnBu")
```

<center><span style="font-size:27px;color:#9ACD32;font-weight:bold">Correlation

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(),linewidths=.9, yticklabels=False,cmap="YlGnBu",square=False,linecolor='black')
```

```python
df.diagnosis.replace(to_replace=['B','M'],value=[0,1],inplace=True)
```

<center><span style="font-size:27px;color:#9ACD32;font-weight:bold">Splitting Data

```python
x= df.drop(['diagnosis','id'],axis =1)
y = df.diagnosis
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=11)
```

<h1 id="heading2">
<span style="font-size:30px;color:#9ACD32;font-weight:bold"><center>Model Fitting</center></span>
<a class="anchor-link" href="https://www.kaggle.com/alampalsingh/notebook004d896e8c#heading2" target ='_self'></a>
</h1>

```python
acc = {}
lr = LogisticRegression()
lr.fit(x_train,y_train)
acc['Logistic Regression'] = accuracy_score(y_test,lr.predict(x_test))
accuracy_score(y_test,lr.predict(x_test))
```

```python
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
acc['K Nearest Neighbors'] = accuracy_score(y_test,knn.predict(x_test))
accuracy_score(y_test,knn.predict(x_test))
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
acc['Decision Tree'] = accuracy_score(y_test,dtc.predict(x_test))
accuracy_score(y_test,dtc.predict(x_test))
```

```python
his = HistGradientBoostingClassifier()
his.fit(x_train,y_train)
acc['Hist Gradient Boosting'] = accuracy_score(y_test,his.predict(x_test))
accuracy_score(y_test,his.predict(x_test))
```

```python
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
acc['Random Forest'] = accuracy_score(y_test,rfc.predict(x_test))
accuracy_score(y_test,rfc.predict(x_test))
```

```python
abc = AdaBoostClassifier()
abc.fit(x_train,y_train)
acc['Ada Boost'] = accuracy_score(y_test,abc.predict(x_test))
accuracy_score(y_test,abc.predict(x_test))
```

```python
xgb = XGBClassifier()
xgb.fit(x_train,y_train)
acc['XGBoost'] = accuracy_score(y_test,xgb.predict(x_test))
accuracy_score(y_test,xgb.predict(x_test))
```

```python
lgb = LGBMClassifier()
lgb.fit(x_train,y_train)
acc['LightGBM'] = accuracy_score(y_test,lgb.predict(x_test))
accuracy_score(y_test,lgb.predict(x_test))
```

```python
key = list(acc.keys())
```

```python
val = [float(acc[k]) for k in key]
```

```python
col = sns.color_palette("YlGnBu_r",8)
col = col.as_hex()
```

```python
fig,acc = plt.subplots(figsize = (16,6))
acc = sns.barplot(val,key,palette="YlGnBu")
acc.set_xlim(0,1.1)
for i in range(0,len(key)):
    acc.text(val[i]-0.05,i,str(np.round(val[i],4)),fontdict = dict(color = col[i],fontsize = 18,ha = 'center'),weight = 'bold')
```

<center><span style="font-size:27px;color:#b7410e;font-weight:bold">Here all the accuracy score we have got may vary with the value of random state. But if we try our every model for prediciton, LightGBM will succeed.
