```python
#имортируем библиотеки для визуализации

import seaborn as sns
import matplotlib.pyplot as plt
```

**Нулевые значения находятся в колонках director, cast,country,date_added и rating.**

Для текущего анализа мы убирем колонки date_added, director и cast. В случае создания рекомендательной системы данная практика неуместна, так как данные по режисёром и актёрам ключевые для рекомендаций пользователям.

```python
netflix.drop(['director', 'cast', 'date_added'], axis = 1, inplace = True)
```

```python
netflix.head()
```

Для всех пустых значений по колонки country используем значение "United States".

```python
netflix['country'].replace(np.nan, 'United States', inplace = True)
```

```python
netflix.head()
```

```python
netflix['rating'].value_counts()
```

```python
netflix['listed_in'].value_counts()
```

10 оставшихся пустых значений мы заменим на самый популярный рейтинг TV-MA.

```python
netflix['rating'].replace(np.nan, 'TV-MA', inplace = True)
```

```python
#проверка

netflix.isnull().sum()
```

```python
#распределение по странам

import plotly.express as px
top_rated = netflix [0:10]
fig = px.sunburst(
    top_rated,
    path = ['country'])
fig.show()
```

```python
#распределение по рейтингу

netflix['rating'].value_counts().plot.pie(autopct = '%1.1f%%', figsize=(10,20))
plt.show()
```

```python
#топ 15 самых старых фильмов от Netflix

old_movies = netflix.sort_values("release_year", ascending = True)
old_movies[["title", "release_year"]][:15]
```

***Распределение контента по странам.***


```python
netflix_countries = pd.DataFrame(netflix.country.value_counts().reset_index().values, columns = ['country', 'count'])
netflix_countries.head()
```

```python
netflix_countries.tail()
```

```python
fig = px.choropleth(
    locationmode = 'country names',
    locations = netflix_countries.country,
    labels = netflix_countries['country'],
)
fig.show()
```

```python
date_netflix = pd.DataFrame(netflix.release_year.value_counts().reset_index().values, columns=['Year', 'Count'])
date_netflix.head()
```

```python
plt.figure(figsize=(12, 6))
netflix[netflix['type']=='Movie']['release_year'].value_counts()[:20].plot(kind='bar', color='Red')
plt.title('Частота фильмов, выпущенных в разные годы и доступных на Netflix')
```

```python
plt.figure(figsize=(12, 6))
netflix[netflix['type']=='TV Show']['release_year'].value_counts()[:20].plot(kind='bar', color='Green')
plt.title('Частота тв-программ, выпущенных в разные годы и доступных на Netflix')
```

```python
plt.figure(figsize=(12, 6))
netflix[netflix['type']=='Movie']['listed_in'].value_counts()[:10].plot(kind='barh', color='Red')
plt.title('Топ 10 жанров фильмов на Netflix', size = 18)
```

```python
plt.figure(figsize=(12, 6))
netflix[netflix['type']=='TV Show']['listed_in'].value_counts()[:10].plot(kind='barh', color='Green')
plt.title('Топ 10 жанров тв-программ на Netflix', size = 18)
```
