```python
bird_data = pd.read_csv("/kaggle/input/birds-bones-and-living-habits/bird.csv")
```

```python
allowed_eco_groups = ['SW','SO','W','T','R','P']

bird_data = bird_data[bird_data["type"].isin(allowed_eco_groups)]

bird_data
```

```python
# Drop empty rows
bird_data.drop(bird_data.index[(bird_data["huml"] == "")],axis=0,inplace=True)
bird_data.drop(bird_data.index[(bird_data["ulnal"] == "")],axis=0,inplace=True)
```

```python
from sklearn.model_selection import train_test_split

train_data, test_data = train_test_split(bird_data, test_size=0.2)
```

```python
features = np.array(train_data[['huml']])
targets = np.array(train_data['type'])

test_features = np.array(train_data[['huml']])
test_targets = np.array(train_data['type'])
```

```python
from sklearn.neighbors import KNeighborsClassifier

# other weights options include 'distance'
knn_model = KNeighborsClassifier(5, weights='uniform')

knn_model.fit(features, targets)
```

```python
import matplotlib.pyplot as plt

labels, index = np.unique(targets, return_inverse=True)
# print(index)
print(labels)

scatter = plt.scatter(features[:,0],features[:,0], c=index)
plt.legend(scatter.legend_elements()[0], labels)
plt.show()
```

```python
predictions = knn_model.predict(test_features)

correct = 0
incorrect = 0

for i in range(len(predictions)):
    print(f"Predicted {predictions[i]}, actually {test_targets[i]}.")
    if predictions[i] == test_targets[i]:
        correct +=1
        print("^Correct")
    else:
        incorrect +=1
        print("^Incorrect")
        
print(f"Got {correct} correct and {incorrect} incorrect.")
print(f"{100*correct/(correct+incorrect)} percent accuracy.")
```
