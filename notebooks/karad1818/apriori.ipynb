```python
df = pd.read_csv("/kaggle/input/groceries-dataset/Groceries_dataset.csv",parse_dates=['Date'])
df.info()
```

```python
# make a item as separate column
item = pd.get_dummies(df.itemDescription)

# delete column item Description
df.drop(columns=['itemDescription'],inplace=True)

df = df.join(item)
```

```python
df.info()

# now aggregate a row with same memeber_number and date
df = df.groupby(["Member_number","Date"]).sum()

df.info()
```

```python
# create a list of column name
c_name = list(df.columns)

# make a list of list from dataframe cause that's what apriori's parameter

data_list = df.values.tolist()

# let's make a list with elements only and remove zero as well
final_list = []
for i in range(0,len(data_list),1):
    temp_list = []
    for j in range(0,len(data_list[i]),1):
        if data_list[i][j] > 0:
            temp_list.append(c_name[j])
    final_list.append(temp_list)
```

```python
# install apriori
!pip install apyori
from apyori import apriori
```

```python
rules = apriori(final_list,min_support=0.00030,min_confidence=0.05,min_lift=3,min_length=2)
for i in rules:
    x = i[0]
    li = [ p for p in x]
    print("Rules : " + li[0] + "  -->  " + li[1])
    print("support : " + str(i[1]))
    print("confidence : " + str(i[2][0][2]))
    print("Lift : " + str(i[2][0][3]))
    print(" -----------------------------------")
```
