```python
#dataframe for ratings
df1 = pd.DataFrame({'total_ratings':df['total ratings']})
df1.reset_index()
df1.index = df['title']
df1
```

```python
#graph illustrating games with highest ratings
df1['total_ratings'][:10].plot(kind='bar')
plt.xlabel('Games')
plt.ylabel('Total Ratings (10M)')
plt.title('Top 10 games with the best ratings')
plt.show()
```

```python
#graph illustrating the games with the least ratings 
df1['total_ratings'][-10:].plot(kind='barh', color = 'purple')
plt.ylabel('Total Ratings')
plt.xlabel('Games')
plt.title('Top 10 Games with least ratings')
plt.show()
```

```python
df['paid'].value_counts()
```

```python
#Price comparison
plt.figure(figsize= (14, 9))

ax= sns.countplot(x= 'price', data= df, order= df['price'].value_counts().index)
plt.title('Price')
for patch in ax.patches:
    h= patch.get_height()
    w= patch.get_width()
    l= patch.get_x()
    p= h/len(df)*100
    ax.text(l + w/2, h + 10,
           '{}\n ({:.2f}%)'.format(h, p), ha= 'center')
```

```python
# Total ratings
df['good_r']= df['4 star ratings'] + df['5 star ratings']
df['bad_r']= df['1 star ratings'] + df['2 star ratings'] + df['3 star ratings']

# Percent of ratings
df['good_r%']= df['good_r'] / (df['good_r'] + df['bad_r'])
df['bad_r%']= df['bad_r'] / (df['good_r'] + df['bad_r'])
```

```python
#calculating top ratings for each game sector
df2= df[df['total ratings'] > 1000000]
df2.sort_values('good_r%', ascending= False).head(10)[['title', 'good_r%', 'total ratings']]
```

```python
plt.figure(figsize= (15, 150))
x = 1
for i in df2['category'].unique():
    plt.subplot(17, 1, x)
    sns.set_palette("pastel")
    plots= df2[df['category'] == i].sort_values('good_r%', ascending= False).head(10)[['title', 'good_r%', 'total ratings']]
    
    ax= sns.barplot(y= 'title', x= 'good_r%', data= plots,)
    plt.title('Top 10 '+i+ ' ratings ')
    plt.xticks(rotation= 75)
    plt.xlabel('')
    plt.ylabel('')
    for patch in ax.patches:
        h= patch.get_height()
        w= patch.get_width()
        b= patch.get_y()
        l= patch.get_x()
    
        ax.text(l + w*1.0285 , b + 0.5,
               '{:.2f}%'.format(w*100), ha= 'center')
    x+= 1
```

```python
#Total number of games
df['category'].value_counts()
plt.figure()
C = pd.value_counts(df['category'].values, sort=True).head(10)
C.plot.barh(color = 'red')
plt.ylabel('Category')
plt.xlabel('No of games')
plt.title('Total No of Games')
plt.show()
```
