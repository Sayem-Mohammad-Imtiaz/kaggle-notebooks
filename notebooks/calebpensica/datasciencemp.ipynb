**1. Which console has sold the most games as of 2016?**

```python
salesRating = df[['Platform', 'Global_Sales', 'Critic_Score']]
platforms = salesRating.Platform.unique()
totalSales = salesRating[['Platform','Global_Sales']]
sumSales = totalSales.groupby(['Platform']).sum()
print(sumSales.sort_values(['Global_Sales'], ascending=False))
```

The result shows that PS2, Xbox360, and PS3 has sold the most games out of all consoles. Is there a correlation with the number of available games per console with the amount sold? 

```python
nPS2 = df[(df.Platform=='PS2')].size
nX360 = df[(df.Platform=='X360')].size
nPS3 = df[(df.Platform=='PS3')].size
nWii = df[(df.Platform=='Wii')].size
nDS = df[(df.Platform=='DS')].size

y = [nPS2, nX360, nPS3, nWii, nDS]
N = len(y) + 2000
x = ["PS2", "Xbox360", "PS3", "Wii", "DS"]
width = 1/1.5
plt.bar(x,y,width,color="green")
```

One can infer that one of the reasons why the PS2 holds the record of the most sold games is because of the volume of games available. However, since the result shows that the DS exceeds even the PS2 in terms of the amount of games, it shows that the volume of games does not always equal to higher quantities sold. 

**2. How does the Critic Score correlate with the total global sales?**

```python
highest_selling = df[np.isfinite(df['Critic_Score'])].head(10)
highest_selling = highest_selling[['Name', 'Platform','Year_of_Release', 'Critic_Score', 'Global_Sales']]
highest_selling = highest_selling.sort_values(['Global_Sales'],ascending=False)
highest_selling
```

The table shows that the highest selling games have average critic scores. What these games have in common, however, are its release dates, in which they range from 2005 to 2010. In order to gain a better understanding, the following table shows the same table information, but for games that are released after 2010. This will check if the output differs for games released prior to 2010.

```python

highest_selling = df[np.isfinite(df['Critic_Score'])].head(100)
highest_selling = highest_selling[['Name', 'Platform','Year_of_Release', 'Critic_Score', 'Global_Sales']]
highest_selling = highest_selling.sort_values(['Global_Sales'],ascending=False)
highest_selling[highest_selling.Year_of_Release>2010].head(10)
```

#We can see that the highest selling games released after 2010 have a relatively higher Critic Score than the highest selling games prior to 2010.
The table shows that the highest selling games globally released after 2010 have a relatively higher critic scores than the highest selling games released before 2010. One can infer that the critic scores started becoming relevant after 2010.
