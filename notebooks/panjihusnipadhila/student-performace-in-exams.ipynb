# 3. Analyze what score they got for each gender

```python
plt.hist(df['math score'].loc[df['gender']=='male'])
plt.show() 
```

```python
plt.hist(df['math score'].loc[df['gender']=='female'])
plt.show() 
```

```python
x = df[['math score']].loc[(df['gender']=='male') & (df['math score'] >= 80)]
y = df[['math score']].loc[(df['gender']=='female') & (df['math score'] >= 80)]
score = pd.DataFrame({'Male' : [len(x)], 'Female': [len(y)]})
score
#plt.hist(score)
#plt.show()
```

```python
A_score_math = df.loc[df['math score'] >= 75]
score = A_score_math[['gender']]
plt.hist(score)
plt.title("Student's gender who success in math exam")
plt.show()
```

```python
A_score_reading = df.loc[df['reading score'] >= 75]
score = A_score_reading[['gender']]
plt.hist(score)
plt.title("Student's gender who success in reading exam")
plt.show()
```

```python
A_score_writing = df.loc[df['writing score'] >= 75]
score = A_score_writing[['gender']]
plt.hist(score)
plt.title("Student's gender who success in writing exam")
plt.show()
```


**It can be seen that men do better on math exams. however women did better on both writing and reading tests**
