<h1 style='background:#212AA5; border:0; color:black'><center><font color="white">Analysis</font></center></h1> 


We will plot the time evolution of the vaccination.


```python
data_df = pd.read_csv("/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv")
```

<small><a href='#0'>Go to top</a></small>  

```python
country_vaccine_time = data_df[["country", "vaccines", "date", 'total_vaccinations', 
                                'total_vaccinations_per_hundred',  'people_vaccinated','people_vaccinated_per_hundred',
                               'daily_vaccinations','daily_vaccinations_per_million', 
                                'people_fully_vaccinated', 'people_fully_vaccinated_per_hundred'
                               ]].dropna()
country_vaccine_time.columns = ["Country", "Vaccines", "Date", 'Total vaccinations', 'Percent', 'People vaccinated', 'People percent',
                               "Daily vaccinations", "Daily vaccinations per million", 
                                'People fully vaccinated', 'People fully vaccinated percent']
```

```python
from plotly.validators.scatter.marker import SymbolValidator

def plot_time_variation_countries_group(data_df, feature, title, countries):
    data = []
    for country in countries:
        df = data_df.loc[data_df.Country==country]
        trace = go.Scatter(
            x = df['Date'],y = df[feature],
            name=country,
            mode = "markers+lines",
            marker_color= 'lightblue',
            line_color= "darkblue",
            marker_line_width = 2,
            marker_size = 9,
            marker_line_color = 'darkblue',
            marker_symbol='hexagram',
            text=df['Country'])
        data.append(trace)
    layout = dict(title = title,
          xaxis = dict(title = 'Date', showticklabels=True,zeroline=True, zerolinewidth=1, zerolinecolor='grey',
                       showline=True, linewidth=2, linecolor='darkblue', mirror=True,
                       tickfont=dict(size=10,color='black'),), 
          yaxis = dict(title = feature, gridcolor='lightblue', zeroline=True, zerolinewidth=1, zerolinecolor='grey',
                       showline=True, linewidth=2, linecolor='darkblue', mirror=True),
                       plot_bgcolor = 'rgba(0, 0, 0, 0)', paper_bgcolor = 'rgba(0, 0, 0, 0)',
         hovermode = 'x', 
         height=500
         )
    fig = dict(data=data, layout=layout)
    iplot(fig, filename='all_countries')
```

```python
countries = ['Israel']
```

```python
plot_time_variation_countries_group(country_vaccine_time, 'Percent', 'Total vaccination percent evolution', countries)
```

```python
plot_time_variation_countries_group(country_vaccine_time, 'Total vaccinations', 'Total vaccination evolution', countries)
```

```python
plot_time_variation_countries_group(country_vaccine_time, 'People percent', 'People vaccinated percent evolution', countries)
```

```python
plot_time_variation_countries_group(country_vaccine_time, 'People vaccinated', 'People vaccinated evolution', countries)
```

```python
plot_time_variation_countries_group(country_vaccine_time, 'Daily vaccinations', 'Daily vaccinations evolution', countries)
```

Daily vaccinations decreased in the last period. The reason is that most of the population that wants to be vaccinated was already vaccinated.

```python
plot_time_variation_countries_group(country_vaccine_time, 'Daily vaccinations per million', 'Daily vaccinations per million evolution', countries)
```

```python
plot_time_variation_countries_group(country_vaccine_time, 'People fully vaccinated percent', 'People fully vaccinated percent evolution', countries)
```

```python
plot_time_variation_countries_group(country_vaccine_time, 'People fully vaccinated', 'People fully vaccinated evolution', countries)
```

<h1 style='background:#212AA5; border:0; color:black'><center><font color="white">Conclusions</font></center></h1> 


A percent of 62.37% (~60%) of the population vaccinated might not seems too much. We should remember though that the targeted population is above 18 years old. But given the current population pyramid in Israel, with:
* 27.89% 0-14 years;
* 59.91% 15-64 years;
* 12.21% 65 and older,
the statistic of population percentage vaccinated should be interpreted differently. Let's consider that target population is above 14 years old. We have then 72.12% of the population. Now, the 59.96% from entire population means actually 83% from entire targeted population. This, even with the more agresive COVID-19 Indian (type D) could still ensure herd immunization.
