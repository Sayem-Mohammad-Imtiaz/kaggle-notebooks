Romania is one of the countries in European Union with the highest percent of vaccination. As a vaccination scheme, they started using Pfizer/BioNTech, the first vaccination scheme that was issued. In the next weeks, two more vaccination schemes were added, Moderna and Oxford/AstraZeneca (limited to people unde 55 years old). From May, also Johnson&Johnson was added to the vaccinations options.

Vaccination is scheduled in Romania in 3 main steps, as following:
- 1st wave, for all medical staff (doctors, nurses, ambulance workers ...);  for them Pfizer/BioNTech vaccines were used; most of them already had the rapel administred (21 days after the first shot); 
- 2nd wave, people over 65 years old, with chronical medical conditions (that increase the likelihood of severe illness if infected with COVID-19) and from essential activities (key government positions, industries essential for fighting the pandemics ...); they also were administred primarly Pfizer/BioNTech, but now they are also getting Moderna and Oxford/AstraZeneca.
- 3rd wave, for general population; this wave is as well divided in two contingents, III-A, with mostly staff from private business areas considered to have priority, due to exposure to public or dependence of other economic areas on them (extractive and manufacturing industries, hairdressing and personal care services, bank clerks and supermarket staff); for them it is reserved Oxford/AstraZeneca; the group III-B represents the general population and them are not yet scheduled.
Update: Currently, based on registering on waiting lists at various vaccination centers, everybody is programmed automatically, without being asked to opt for a certain vaccine, with either Pfizer, Moderna or Oxford/AstraZeneca. Update: Johnson&Johnson is also an option.

After mid-April, vaccination programme became more diverse, including:
- Vaccination marathons over weekend, first one being held in Timisoara, one of the large cities in W of Romania;
- Vaccination drive-through in several large cities, including the Capital city, Bucharest;
- Vaccination without reservation (first arrived get a jab) in several large cities.

The most difficult problem for the authorities will be to reach to the small communities, in the countryside and small towns, where more than 40% of the Romanian population is still living.

Population acceptance is also an important factor. Romanian President was the first person to be vaccinated in 2nd wave and the procedure was performed in from of television cameras.

Note June 22: In recent weeks, the dynamic of vaccination slowed down.

We will plot the time evolution of the vaccination.


```python
data_df = pd.read_csv("/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv")
data_manufacturer_df = pd.read_csv("/kaggle/input/covid-world-vaccination-progress/country_vaccinations_by_manufacturer.csv")
```

```python
data_romania_df = data_manufacturer_df.loc[data_manufacturer_df.location=="Romania"]
manufacturers = data_romania_df.vaccine.unique()
```

```python
def plot_time_variation_manufacturer(data_romania_df, feature, title, manufacturers, y_type=None):
    data = []
    for manufacturer in manufacturers:
        df = data_romania_df.loc[data_romania_df.vaccine==manufacturer]
        trace = go.Scatter(
            x = df['date'],y = df[feature],
            name=manufacturer,
            mode = "lines",
            marker_line_width = 2,
            text=df['vaccine'])
        data.append(trace)
    layout = dict(title = title,
          xaxis = dict(title = 'date', showticklabels=True,zeroline=True, zerolinewidth=1, zerolinecolor='grey',
                       showline=True, linewidth=2, linecolor='darkblue', mirror=True,
                       tickfont=dict(size=10,color='black'),), 
          yaxis = dict(title = feature, gridcolor='lightblue', zeroline=True, zerolinewidth=1, zerolinecolor='grey',
                       showline=True, linewidth=2, linecolor='darkblue', mirror=True, type=y_type),
                       plot_bgcolor = 'rgba(0, 0, 0, 0)', paper_bgcolor = 'rgba(0, 0, 0, 0)',
         hovermode = 'x', 
         height=500
         )
    
    fig = dict(data=data, layout=layout)
    iplot(fig, filename='all_vaccines')
```

```python
plot_time_variation_manufacturer(data_romania_df, 'total_vaccinations', 'Total vaccinations per manufacturer', manufacturers)
```

```python
plot_time_variation_manufacturer(data_romania_df, 'total_vaccinations', 'Total vaccinations per manufacturer (log scale)', manufacturers, "log")
```

In Dec, 27, 2020, Romania started to vaccinate with Pfizer/BioNTech.
Feb, 4, 2021, started vaccination with Moderna.
Feb, 15 started innoculating as well with Oxford/AstraZeneca.
on May, 3 also Johnson&Johnson started to be used.

On May, 12, Romania passed threshold of 5M doses of Pfizer/BioNTech administrated.

```python
country_vaccine_time = data_df[["country", "vaccines", "date", 'total_vaccinations', 
                                'total_vaccinations_per_hundred',  'people_vaccinated','people_vaccinated_per_hundred',
                               'daily_vaccinations','daily_vaccinations_per_million', 
                                'people_fully_vaccinated', 'people_fully_vaccinated_per_hundred'
                               ]].dropna()
country_vaccine_time.columns = ["Country", "Vaccines", "Date", 'Total vaccinations', 'Percent', 'People vaccinated', 'People percent',
                               "Daily vaccinations", "Daily vaccinations per million", 
                                'People fully vaccinated', 'People fully vaccinated percent']
```

```python
def plot_time_variation_countries_group(data_df, feature, title, countries):
    data = []
    for country in countries:
        df = data_df.loc[data_df.Country==country]
        trace = go.Scatter(
            x = df['Date'],y = df[feature],
            name=country,
            mode = "lines",
            marker_color= 'yellow',
            marker_line_width = 2,
            marker_size = 9,
            marker_line_color = 'blue',
            line_color = "red",
            text=df['Country'])
        data.append(trace)
    layout = dict(title = title,
          xaxis = dict(title = 'Date', showticklabels=True,zeroline=True, zerolinewidth=1, zerolinecolor='grey',
                       showline=True, linewidth=2, linecolor='darkblue', mirror=True,
                       tickfont=dict(size=10,color='black'),), 
          yaxis = dict(title = feature, gridcolor='lightblue', zeroline=True, zerolinewidth=1, zerolinecolor='grey',
                       showline=True, linewidth=2, linecolor='darkblue', mirror=True),
                       plot_bgcolor = 'rgba(0, 0, 0, 0)', paper_bgcolor = 'rgba(0, 0, 0, 0)',
         hovermode = 'x', 
         height=500
         )
    fig = dict(data=data, layout=layout)
    iplot(fig, filename='all_countries')
```

```python
countries = ['Romania']
```

```python
plot_time_variation_countries_group(country_vaccine_time, 'Percent', 'Total vaccination percent evolution', countries)
```

```python
plot_time_variation_countries_group(country_vaccine_time, 'Total vaccinations', 'Total vaccination evolution', countries)
```

```python
plot_time_variation_countries_group(country_vaccine_time, 'People percent', 'People vaccinated percent evolution', countries)
```

```python
plot_time_variation_countries_group(country_vaccine_time, 'People vaccinated', 'People vaccinated evolution', countries)
```

On April, 26 Romania passed the threshold of 3M people vaccinated.

```python
plot_time_variation_countries_group(country_vaccine_time, 'Daily vaccinations', 'Daily vaccinations evolution', countries)
```

First time to pass the threshold of 100K vaccinated per day was May, 11.

```python
plot_time_variation_countries_group(country_vaccine_time, 'Daily vaccinations per million', 'Daily vaccinations per million evolution', countries)
```

Daily vaccination evolution is, as of June 22 2021, decreasing from a peak of 106K vaccines / day on May 15.

```python
plot_time_variation_countries_group(country_vaccine_time, 'People fully vaccinated', 'People fully vaccinated evolution', countries)
```

First time to pass the threshold of 2.5M people fully vaccinated was May, 12.

```python
plot_time_variation_countries_group(country_vaccine_time, 'People fully vaccinated percent', 'People fully vaccinated percent evolution', countries)
```

Currently (as of Sept 1st), rate of daily vaccinations decreased drastically and percent of total population vaccinated saturated to ~27%.
