## Identify Interesting Statistics

This dataset is already cleaned for us and relatively straightforward. In this dataset, there doesn't have any missing data or tricky values. 

To do some statistics, I will 'import numpy' first.

```python
import numpy as np
```

```python
# The total number of entries in the dataset
print(len(df))
```

```python
# Using different way to find maximum、minimum and average
# The max、 min、 range、 mean of Women Entrepreneurship
women_entre_max = np.max(df["Women Entrepreneurship Index"])
women_entre_min = np.min(df["Women Entrepreneurship Index"])
women_entre_range = women_entre_max - women_entre_min
women_entre_mean = np.mean(df["Women Entrepreneurship Index"])
print("Women Entrepreneurship:")
print("Max:", women_entre_max, "Min:", women_entre_min, "Range:", women_entre_range, "Mean:", women_entre_mean)

# The max、 min、 range、 mean of Entrepreneurship
entre_max = np.max(df["Entrepreneurship Index"])
entre_min = np.min(df["Entrepreneurship Index"])
entre_range = entre_max - entre_min
entre_mean = np.mean(df["Entrepreneurship Index"])
print("Entrepreneurship:")
print("Max:", entre_max, "Min:", entre_min, "Range:", entre_range, "Mean:", entre_mean)

# The max、 min、 range、 mean of Female Labor Force Participation Rate
female_partic_rate_max = np.max(df["Female Labor Force Participation Rate"])
female_partic_rate_min = np.min(df["Female Labor Force Participation Rate"])
female_partic_rate_range = female_partic_rate_max - female_partic_rate_min
female_partic_rate_mean = np.mean(df["Female Labor Force Participation Rate"])
print("Female Labor Force Participation Rate")
print("Max:", female_partic_rate_max, "Min:", female_partic_rate_min, "Range:", female_partic_rate_range, "Mean:", female_partic_rate_mean)
```

### Differences of Female Labor Force Participation Rate between different Level of development

```python
# Difference averages in the Female Labor Force Participation Rate between different Level of development
# Filter the data which is developed or developing

dev_ed = df[df["Level of development"]=="Developed"]
dev_ing = df[df["Level of development"]=="Developing"]

dev_ing.reset_index(inplace = True)
```

```python
# The average of Female Labor Force Participation Rate in Develped country
female_partic_rate_dev_ed = np.mean(dev_ed["Female Labor Force Participation Rate"])
# The average of Female Labor Force Participation Rate in Develping country
female_partic_rate_dev_ing = np.mean(dev_ing["Female Labor Force Participation Rate"])

print('The mean of Female Labor Force Participation Rate in Develped country is', female_partic_rate_dev_ed)
print('The mean of Female Labor Force Participation Rate in Develping country is', female_partic_rate_dev_ing)
```

```python
# Difference max、min in the Female Labor Force Participation Rate between different Level of development
# The Female Labor Force Participation Rate in Develped country
female_partic_rate_dev_ed_max = np.max(dev_ed["Female Labor Force Participation Rate"])
female_partic_rate_dev_ed_min = np.min(dev_ed["Female Labor Force Participation Rate"])
# The Female Labor Force Participation Rate in Develping country
female_partic_rate_dev_ing_max = np.max(dev_ing["Female Labor Force Participation Rate"])
female_partic_rate_dev_ing_min = np.min(dev_ing["Female Labor Force Participation Rate"])

# Difference range in the Female Labor Force Participation Rate between different Level of development
# The Female Labor Force Participation Rate in Develped country
female_partic_rate_dev_ed_range = female_partic_rate_dev_ed_max - female_partic_rate_dev_ed_min
# The Female Labor Force Participation Rate in Develping country
female_partic_rate_dev_ing_range = female_partic_rate_dev_ing_max - female_partic_rate_dev_ing_min

print('The maximum/minimum of Female Labor Force Participation Rate in Develped country is',
     female_partic_rate_dev_ed_max, "/", female_partic_rate_dev_ed_min)
print('The maximum/minimum of Female Labor Force Participation Rate in Develping country is',
     female_partic_rate_dev_ing_max, "/", female_partic_rate_dev_ing_min)

print('The range of Female Labor Force Participation Rate in Develped country is', female_partic_rate_dev_ed_range)
print('The range of Female Labor Force Participation Rate in Develping country is', female_partic_rate_dev_ing_range)
```

### Box Plots

#### Woman Entrepreneurship for different Level of Development / Entrepreneurship for different Level of Development

```python
# Using subplot to plot two plots in one way
plt.figure(figsize=(15,15))
plt.subplot(221)
sns.boxplot(data = df, x = 'Level of development', y = 'Women Entrepreneurship Index')
plt.title("Women Entrepreneurship Index")
 
plt.subplot(222)
sns.boxplot(data = df, x = 'Level of development', y = 'Entrepreneurship Index')
plt.title("Entrepreneurship Index")

plt.show()
```

#### Female Labor Force Participation Rate for different Level of Development

```python
plt.figure(figsize=(10,6))
sns.boxplot(data = df, x = 'Level of development', y = 'Female Labor Force Participation Rate')
```

Developed Countries have more Female Labor Participation than Developing Countries. Besides, we can also find that the range of Female Labor Participation in Developed Countries is smaller than Developing Countries.

### Map

#### Using Plotly to plot the world map

#### Woman Entrepreneurship Indices in the World

```python
# Not Learn from this course, but it is suitable for this dataset
# Country name : ISO-3 version (search from the Net)

# plotly.express choropleth locations/locationmode need to use ‘ISO-3’, ‘USA-states’, or ‘country names’ .
# First need to search the countries' iso3 name.
country = pd.read_csv("../input/iso3-country/ISO3.csv", encoding='utf-8')
country.head()
```

```python
# Merge the data together
df_map = pd.merge(df, country, on = 'Country')
df_map.head()
```

```python
fig = px.choropleth(df_map, locations='iso3', color = 'Women Entrepreneurship Index', hover_name='Country',
                    hover_data = ['Level of development', 'Entrepreneurship Index', 'Women Entrepreneurship Index',
                                  'Female Labor Force Participation Rate'],
                    color_continuous_scale=px.colors.sequential.Viridis, title="Woman Entrepreneurship Index in the World")
fig.show()
```

As the world map shown above, we can find that Women Entrepreneurship is higher in Europe. Maybe this is true that women entrepreneurs are exactly more than other countries, but there are some countries not contained in this dataset.
