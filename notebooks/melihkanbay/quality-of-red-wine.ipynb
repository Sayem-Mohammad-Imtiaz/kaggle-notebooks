I examined the factors of the quality of red wine with correlation technics.
What is correlation?
Correlation is a mutual relationship or connection between two or more things.
A correlation coefficient is a way to put a value to the relationship. Correlation coefficients have a value of between -1 and 1. A “0” means there is no relationship between the variables at all, while -1 or 1 means that there is a perfect negative or positive correlation.

The following is a correlation map. we can see the correlation coefficients on the map.

```python
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```

If the numbers are high, it is useful.  
Alcohol, volatile acidity, sulphates and citric acid is useful parameters for us.

The first graph(effect of alcohol on wine quality) shows the positive correlation of alcohol and quality.   
its means: If the alcohol is high, the quality of the wine is will be high.

```python
fig = plt.figure(figsize = (18,6))
sns.barplot(x = 'quality', y = 'alcohol', data = data)
plt.title("Effect of alcohol on wine quality")
plt.show()
```

Second graph is Effect of sulphates on wine quality.   
its positive correlation


```python
fig = plt.figure(figsize = (18,6))
sns.barplot(x = 'quality', y = 'sulphates', data = data)
plt.title("Effect of sulphates on wine quality")
plt.show()
```

Third graph is Effect of citric acid on wine quality.   
Its positive correlation

```python
fig = plt.figure(figsize = (18,6))
sns.barplot(x = 'quality', y = 'citric acid', data = data)
plt.title("Effect of citric acid on wine quality")
plt.show()
```

Fourth graph is Effect of volatile acidty on wine quality.   
Its negative correlation

```python
fig = plt.figure(figsize = (18,6))
sns.barplot(x = 'quality', y = 'volatile acidity', data = data)
plt.title("Effect of volatile acidity on wine quality")
plt.show()
```
