```python
data.Attack.plot(color="red",figsize = (22,12))
data.Defense.plot(color="green",figsize = (22,12))
plt.legend()
plt.title("Attack and Defense line")
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.show()
```

```python
data.plot(kind="scatter", x="Attack",y="Defense",color="red",alpha=0.6,figsize = (12,10))
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.title("Attack and Defense Correlation")
plt.show()
```

```python
data.Speed.plot(kind = 'hist',bins = 70,figsize = (12,12))
plt.title("Frequency of speed")
plt.legend()
plt.show
```

```python
data = pd.read_csv("../input/pokemon-challenge/pokemon.csv")
```

**Defensive heroes, they have +200 Defense**

**Offensive heroes, they have  +175 Attack**

```python
offensive_heroes=data["Attack"]>175
data[offensive_heroes]
```

**Perfect heroes, they have  +200 Defense and +100 Attack**

```python
data[np.logical_and(data['Defense']>200, data['Attack']>100 )]
```
