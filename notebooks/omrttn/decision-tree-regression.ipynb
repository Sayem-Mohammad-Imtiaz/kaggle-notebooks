# 1-Decision Tree Regression

*In this model, a value of y dependent on x1 and x2 is obtained. Each splited region shown in figure-1 is named terminal leaf. Firstly, data is splited according to information entropy formula and than prediction is made. An example of the split process is given in figure 2. 
In the example given in Figure 2, the parts with constant numbers were found by taking the average of the y values in that region. In this figure the average y values are randomly given.*


* Import Library

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeRegressor
```

* Import and plot data

```python
df = pd.read_csv("../input/biomechanical-features-of-orthopedic-patients/column_2C_weka.csv")

plt.scatter(df.lumbar_lordosis_angle,df.sacral_slope)
plt.xlabel("lumbar lordosis")
plt.ylabel("sacral slope")
plt.show()
```

* Define x and y

```python
x = df.iloc[:,2].values.reshape(-1,1)
y = df.iloc[:,3].values.reshape(-1,1)
```

* Fit and prediction

```python
dtree_reg = DecisionTreeRegressor()
dtree_reg.fit(x,y)

prediction = dtree_reg.predict([[30]])
print ("prediction: ",prediction)
```

* Visualize

```python
x_ = np.arange(min(x),max(x),10).reshape(-1,1)
y_head = dtree_reg.predict(x_)
plt.scatter(x,y,color="red")
plt.plot(x_,y_head,color = "green")
plt.xlabel("lumbar lordosis")
plt.ylabel("sacral slope")
plt.show()
```
