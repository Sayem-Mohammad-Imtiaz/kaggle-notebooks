```python
sorted_corruption= df.sort_values("Perceptions of corruption")
```

```python
plt.figure(figsize=(10,6))
sns.barplot(sorted_corruption["Perceptions of corruption"][:10], sorted_corruption["Country or region"][:10]);
plt.title("The 10 countries with the lowest perception of corruption");
```

```python
plt.figure(figsize=(10,6))
sns.barplot(sorted_corruption["Country or region"][-10:], sorted_corruption["Perceptions of corruption"][-10:]);
plt.xticks(rotation= 45);
plt.title("The 10 countries with the highest perception of corruption");
```

```python
# 10 countries with low corruption perception --> GDP per capita vs Social support vs score
f,ax = plt.subplots(figsize = (15,9))
sns.barplot(y=sorted_corruption[:10]["GDP per capita"] ,x=sorted_corruption[:10]["Country or region"], 
            color='red',alpha = .8, label='GDP per capita' )

sns.barplot(y=sorted_corruption[:10]["Social support"] ,x=sorted_corruption[:10]["Country or region"], 
            color='yellow',alpha = .6, label='Social support')

sns.barplot(y=sorted_corruption[:10]["Score"] ,x=sorted_corruption[:10]["Country or region"], 
            color='lime',alpha = .4, label='Score')

ax.legend(loc='lower right',frameon = True) 
plt.title("GDP per capita vs Social support vs score")
plt.xticks(rotation= 45);
```

```python
# 10 countries with high corruption perception --> GDP per capita vs Social support vs score
f,ax = plt.subplots(figsize = (15,9))
sns.barplot(y=sorted_corruption[-10:]["GDP per capita"] ,x=sorted_corruption[-10:]["Country or region"], 
            color='red',alpha = .8, label='GDP per capita' )

sns.barplot(y=sorted_corruption[-10:]["Social support"] ,x=sorted_corruption[-10:]["Country or region"], 
            color='yellow',alpha = .6, label='Social support')

sns.barplot(y=sorted_corruption[-10:]["Score"] ,x=sorted_corruption[-10:]["Country or region"], 
            color='lime',alpha = .4, label='Score')

ax.legend(loc='lower right',frameon = True)
plt.title("GDP per capita vs Social support vs score")
plt.xticks(rotation= 45);
```
