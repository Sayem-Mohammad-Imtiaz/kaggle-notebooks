### 3.1 Countries with Most Number of Titles

We find the frequency of top 5 values for column country after dropping the missing values.

```python
df['country'].dropna(inplace = True)
print(dict(Counter(df['country'].values).most_common(5)))
```

### 3.2 Movie Duration

To analyze the movie duration, firstly we'll filter the dataset to include only movies and then print a set of durations of all the movies.

```python
df_movie = df[df['type'] =='Movie']
print(set(df_movie['duration']))
```

As we can see all the durations are of the type 'string' and have 'min' written after them. So, we'll remove the 'min' and typecast every observation as 'int' 

```python
df_movie['duration'] = df_movie['duration'].map(lambda d: d.rstrip('min')).astype(int)
print(set(df_movie['duration']))
```

Now that we have the durations, we'll visualize them against their genre using a boxplot. But, since there are so many genres we'll use the top 5 most frequent ones.
The Boxplot visualizes the data based on numeric values like minimum, maximum, median, first quartile, and third quartile.

```python
keys = []
for i in dict(Counter(df_movie['listed_in'].values).most_common(5)):
    keys.append(i)
print(keys)
df_new = df_movie[df_movie['listed_in'].isin(keys)]
sns.boxplot(x = df_new['listed_in'], y = df_new['duration'])
```

We can also visualize the duration using a Histogram to show the distribution.

```python
df_movie['duration'].hist(bins=100)
```
