## Kmeans Cluster

 -Objective of K-means is  group similar data points together and discover underlying patterns 
 
 -To achieve this objective, K-means looks for a fixed number (k) of clusters in a dataset so K indicate number of cluster.
 
 -A cluster refers to a collection of data points aggregated together because of certain similarities.
 To find best K value for our data,I am going to use elbow method.


```python
wcss=[]# We will try to minimaze wcss 

for K in range(1,15):
    
    Kmeans = KMeans(n_clusters=K)
    Kmeans.fit(data)
    wcss.append(Kmeans.inertia_) # inertia_ calculates wcss.
    
plt.plot(range(1,15),wcss)
plt.xlabel("Number of K values")
plt.ylabel("Wcss")
plt.grid()
    
    
```

Optimum K value is 2

```python
n_Kmeans = KMeans(n_clusters = 2)
clusters = n_Kmeans.fit_predict(data)
df["Kmeans_Label"]= clusters
```

```python
df.head()
```

As you see There are some mistakes in my data so I want to measure sucess of my label accuracy

```python
df.shape #Data has 768 rows,I am going to use this to measure label accuracy
```

```python
accuracy=[]

for i in range(df.shape[0]):
    if df.Outcome[i]==df.Kmeans_Label[i]:
        accuracy.append(df.Kmeans_Label[i])
        
    
    
    
```

## Hierarchical Clustring


```python
from sklearn.preprocessing import normalize # I want to normalize my data so in this way My clustering result might be better than Kmeans Clustering
data_scaled = normalize(data)
data2= pd.DataFrame(data_scaled, columns=data.columns)
data2.head()
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram

merg = linkage(data2,method="ward")
dendrogram(merg,leaf_rotation=90)

plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.axhline(y=6, color='r', linestyle='--')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters=2,affinity='euclidean', linkage='ward')  
df["Hierarchical_Label"]=cluster.fit_predict(data2)
```

```python
df.head()
```

```python
plt.scatter(x="Insulin",y="Glucose",data=df,c="Outcome") # original
```

```python
plt.scatter(x="Insulin",y="Glucose",data=df,c="Hierarchical_Label") 
plt.ylabel("Glucose")
```

```python
accuracy=[]

for i in range(df.shape[0]):
    if df.Outcome[i]==df.Hierarchical_Label[i]:
        accuracy.append(df.Hierarchical_Label[i])
        
```

```python
print("Accuracy:",len(accuracy)/df.shape[0]) 
```
