Koostame tabeli mängijatest selliselt, et nad on järjestatud punktisumma alusel.

```python
(df[["Name", "Games Played", "MIN", "PTS"]]
 .sort_values("PTS", ascending=False))
```

Kokkuvõttev tabel, mis näitab kui palju punkte kogusid mängijad, kes mängisid sama arvu mänge ja kui palju oli neil keskmiselt punkte.

```python
df.groupby("Games Played").aggregate({"MIN": ["sum", "mean",],
                                      "PTS" : ["sum", "mean"]})
```

Histogramm mängijate vanuste kohta.

```python
df.Age.plot.hist(rwidth=0.95,color= "c");
```

Scatterplot, millelt näeb mängijate osaletud mängude arvu ja nende kogutud punktide suhet.

```python
df.plot.scatter("Games Played", "PTS", alpha=0.3, color= "m");
```
