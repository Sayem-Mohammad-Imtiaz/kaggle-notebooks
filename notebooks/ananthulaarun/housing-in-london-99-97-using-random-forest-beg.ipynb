```python
house_mean = df['houses_sold'].mean()        
noc_mean = df['no_of_crimes'].mean()

df = df.fillna({'houses_sold' : house_mean})
df = df.fillna({'no_of_crimes' : noc_mean})
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

area = LabelEncoder()
code = LabelEncoder()

df['area'] = area.fit_transform(df['area'])
df['code'] = code.fit_transform(df['code'])
```

```python
df.head()
```

```python
df=df.drop(['date'],axis=1)
```

```python
x = df.drop(['borough_flag'],axis=1)
y = df['borough_flag']
```

```python

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3)
```

```python
print(x_train.shape)
print(y_train.shape)
```

```python
from imblearn.over_sampling import SMOTE

smt = SMOTE()
x_train, y_train = smt.fit_sample(x_train, y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier 
```

```python
rand= RandomForestClassifier(n_estimators=50)
```

```python
model=rand.fit(x_train,y_train)
```

```python
y_pred_train = rand.predict(x_train)
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_train,y_pred_train))
```

```python
y_pred_test= rand.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred_test))
```
