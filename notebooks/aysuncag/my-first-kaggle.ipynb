```python
temp = list()
clean_data = df.dropna()
clean_data.reset_index(inplace=True)
for ind, element in clean_data.iterrows():
    type_show = element['type']
    country=element['country']
    for director in str(element['director']).split(','):
        temp.append([type_show, director,country])
director_data = pd.DataFrame(temp, columns= ['type', 'director','country'])
director_turkish_data=director_data[director_data.country=='Turkey']
director_turkish_data
```

```python
director_turkish_data_count = director_turkish_data.value_counts().to_frame()
director_turkish_data_count.reset_index(level=[0,1], inplace=True)
famous_turkish_director = director_turkish_data_count.rename(columns={0:'count'})
famous_turkish_director = pd.concat([famous_turkish_director[famous_turkish_director['type'] == 'Movie'], famous_turkish_director[famous_turkish_director['type'] == 'TV Show']])
famous_turkish_director.reset_index(inplace=True)
famous_turkish_director
```

```python
for unique_type in famous_turkish_director['type'].unique():
    bar, ax = plt.subplots(figsize=(10,10))
    sns.barplot(x = 'director', y = 'count', data = famous_turkish_director[famous_turkish_director['type'] == unique_type].iloc[:5])
    plt.xlabel('Director in {}'.format(str(unique_type)))
    plt.ylabel('Frequency')
    plt.title('Famous Director in {}'.format(str(unique_type)), size=20)
```
