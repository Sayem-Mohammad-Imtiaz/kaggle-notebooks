# Truco 93: Combinar las categorías pequeñas en una sola categoría llamada "Otros" (usando frecuencias)
[Regresar a tabla de contenido](#table_of_contents)

```python
d = {"genre": ["A", "A", "A", "A", "A", "B", "B", "C", "D", "E", "F"]}
df = pd.DataFrame(d)
df

# Paso 1: Contar la frecuencias
frequencies = df["genre"].value_counts(normalize = True)
frequencies

# Paso 2: Establecer su umbral y filtrar las categorias mas pequeñas
threshold = 0.1
small_categories = frequencies[frequencies < threshold].index
small_categories

# Paso 3: reemplazar los valores
df["genre"] = df["genre"].replace(small_categories, "Other")
df["genre"].value_counts(normalize = True)
```

# Trick 86: Named aggregations - avoids multiindex
[Go back to the Table of Contents](#table_of_contents)

```python
print_files()

df = pd.read_csv("/kaggle/input/titanic/train.csv")
df.head()

# Problema 1
print("El problema se basa en que no sabemos el nombre de la columna")
df.groupby("Pclass")["Age"].agg(["mean", "max"])

# Problema 2
print("El problema se basa en que tenemos multiindex")
df.groupby("Pclass").agg({"Age":["mean", "max"]})

# Nueva solucion con pandas 0.25 o version mas alta
print("Ahora hemos resuelto los problemas anteriores especificando el nombre final de la columna que queremos.")
print("PERO SÓLO FUNCIONA CON UNA COLUMNA. A ESTE TIPO DE OPERACIONES EN VARIAS COLUMNAS COMPRUEBE LA SIGUIENTE CELDA")
df.groupby("Pclass")["Age"].agg(age_mean = "mean", age_max = "max")
```

# Trick 86bis: Agregaciones con nombre en varias columnas: evitando multiindex
[Regresar a la tabla de contenido](#table_of_contents)

```python
def my_agg(x):
    names = {
        'age_media': x['Age'].mean(),
        'age_max':  x['Age'].max(),
        'fare_mean': x['Fare'].mean(),
        'fare_max': x['Fare'].max()
    } # definir los nombres de columnas personalizadas y las operaciones

    return pd.Series(names, index=[ key for key in names.keys()]) # todas las columnas que cree en el diccionario anterior estarán en esta lista de comprensión

df.groupby('Pclass').apply(my_agg)

# Referencia
# https://stackoverflow.com/questions/44635626/rename-result-columns-from-pandas-aggregation-futurewarning-using-a-dict-with
```

# Truco 85: Convertir un tipo de valores a otros
[Go back to the Table of Contents](#table_of_contents)

```python
# Do some fast feature eng on the DF
d = {"gender":["male", "female", "male"], "color":["red", "green", "blue"], "age":[25, 30, 15]}
df = pd.DataFrame(d)
df

# Solution
map_dict = {"male":"M", "female":"F"}
df["gender_mapped"] = df["gender"].map(map_dict) # utilizando diccionarios para mapear valores
df["color_factorized"] = df["color"].factorize()[0] # uso de factorize: devuelte matrices (matriz([0, 1, 2]), Index(['rojo', 'verde', 'azul'], dtype='object')) que hemos seleccionado [0]
df["age_compared_boolean"] = df["age"] < 18 # return a True False boolean value

df
```

# Truco 66: Crear un montón de nuevas columnas usando un bucle for y f-strings df[f'{col}_new']

```python
d = {"state":["ny", "CA", "Tx", "FI"], "country":["USA", "usa", "UsA", "uSa"], "pop":[1000000, 2000000, 30000, 40000]}
df = pd.DataFrame(d)
df

int_types = ["int64"]
# creando nuevas columnas
for col in df.columns:
    ctype = str(df[col].dtype)
    if ctype in int_types:
        df[f'{col}_millions'] = df[col]/1000000
    elif ctype == "object":
        df[f'{col}_new'] = df[col].str.upper()
        # también puede soltar las columnas
        df.drop(col, inplace = True, axis = "columns")
        
df
```

# Truco 63: Calcular el recuento de corrida con grupos usando cumcount() + 1

```python
d = {"salesperson":["Nico", "Carlos", "Juan", "Nico", "Nico", "Juan", "Maria", "Carlos"], "item":["Car", "Truck", "Car", "Truck", "cAr", "Car", "Truck", "Moto"]}
df = pd.DataFrame(d)
df

# Corrigiendo las columnas
df["salesperson"] = df["salesperson"].str.title()
df["item"] = df["item"].str.title()

df["count_by_person"] = df.groupby("salesperson").cumcount() + 1
df["count_by_item"] = df.groupby("item").cumcount() + 1
df["count_by_both"] = df.groupby(["salesperson","item"]).cumcount() + 1
df
```

# Truco 55: Filtrar un df con múltiples criterios usando reducir
[Regresar a la tabla de contenido](#table_of_contents)

```python
df = pd.read_csv("/kaggle/input/drinks-by-country/drinksbycountry.csv")
df

print("Filtro clásico difícil de leer y mantener.")
df[(df["continent"] == "Europe") & (df["beer_servings"] > 150) & (df["wine_servings"] > 50) & (df["spirit_servings"] < 60)]

print("Puede dividirlo en varias líneas para que sea más legible. Pero todavía es difícil de leer.")
df[
    (df["continent"] == "Europe") & 
    (df["beer_servings"] > 150) & 
    (df["wine_servings"] > 50) & 
    (df["spirit_servings"] < 60)
]

print("Criterios de ahorro de soluciones como objetos")

cr1 = df["continent"] == "Europe"
cr2 = df["beer_servings"] > 150
cr3 = df["wine_servings"] > 50
cr4 = df["spirit_servings"] < 60

df[cr1 & cr2 & cr3 & cr4]

print(" Solucion sin reducir")
from functools import reduce

# crea nuestros criterios usando lambda
# lambda toma 2 parametros, x y y
# reducir los combina y por cada cr en el (cr1, cr2, cr3, cr4)
criteria = reduce(lambda x, y: x & y, (cr1, cr2, cr3, cr4))
df[criteria]
```
