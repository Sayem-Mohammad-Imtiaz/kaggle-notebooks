```python
#For data processing and visualization
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

#For Clustering Analysis
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn import tree

#For Prediction Modelling
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
data = pd.read_csv("/kaggle/input/crop-recommendation-dataset/Crop_recommendation.csv")
display(data.head())
display(data.tail())
print('Data Shape: ', data.shape)
display(data.isnull().sum())
```

Average value for each Parameters//

Rata-rata nilai tiap parameter

```python
rata = {'Kadar N': ['{:.2f}'.format(data['N'].mean())],
         'Kadar P': ['{:.2f}'.format(data['P'].mean())],
         'Kadar K': ['{:.2f}'.format(data['K'].mean())],
         'Temperatur (C)': ['{:.2f}'.format(data['temperature'].mean())],
         'Kelembaban (%)': ['{:.2f}'.format(data['humidity'].mean())],
         'pH': ['{:.2f}'.format(data['ph'].mean())],
         'Curah Hujan (mm)': ['{:.2f}'.format(data['rainfall'].mean())]}

rata = pd.DataFrame(rata, index=['Rata-Rata'])
display(rata)
```

Crop classification based on each conditions//

Klasifikasi tiap tanaman berdasarkan kondisi yang disebutkan

```python
a = data[data['N'] > 120]['label'].unique()
b = data[data['P'] > 100]['label'].unique()
c = data[data['K'] > 200]['label'].unique()
d = data[data['temperature'] > 30]['label'].unique()
e = data[data['temperature'] < 10]['label'].unique()
f = data[data['humidity'] < 20]['label'].unique()
g = data[data['ph'] > 8]['label'].unique()
h = data[data['ph'] < 4]['label'].unique()
i = data[data['rainfall'] > 200]['label'].unique()
j = data[data['rainfall'] < 40]['label'].unique()

TanamanDetil = pd.DataFrame({'High Nitrogen':pd.Series(a),'High Phospor':pd.Series(b),
                             'High Potassium':pd.Series(c),'High Temperature':pd.Series(d),
                             'Low Temperature':pd.Series(e),'Low Humidity':pd.Series(f),
                             'High pH':pd.Series(g),'Low pH':pd.Series(h),
                             'High Rainfall':pd.Series(i),'Low Rainfall':pd.Series(j)})

TanamanDetil = TanamanDetil.replace(np.nan, '-')

display(TanamanDetil)
```

```python
k=(data[(data['temperature'] < 20) & (data['humidity'] < 40)]['label'].unique())
l=(data[(data['temperature'] > 28) & (data['humidity'] > 40)]['label'].unique())
m=(data[(data['rainfall'] > 200)] ['label'].unique())

TanamanMusim = pd.DataFrame({'Winter Crop':pd.Series(k),'Summer Crop':pd.Series(l),'Rainfall Crop':pd.Series(m)})

TanamanMusim = TanamanMusim.replace(np.nan, '-')

display(TanamanMusim)
```

```python
x = data.drop(['label'], axis=1)
y = data['label']
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.5, random_state=42)
```

```python
#Support Vector Machine Classifier
svcs = SVC()
svcs.fit(xtrain,ytrain)
ypred2 = svcs.predict(xtest)
print("Accuracy Score of svcs:",accuracy_score(ytest,ypred2))

#Decission Tree Classifier
dtcs = DecisionTreeClassifier(criterion='entropy',random_state=0)
dtcs.fit(xtrain,ytrain)
ypred3 = dtcs.predict(xtest)
print("Accuracy Score of dtcs:",accuracy_score(ytest,ypred3))

#Random Forest Classifier
rfcs = RandomForestClassifier(random_state=0)
rfcs.fit(xtrain, ytrain)
ypred4 = rfcs.predict(xtest)
print("Accuracy Score of rfcs:",accuracy_score(ytest,ypred4))
```

Karena yang paling besar RFCS kita pakai itu.

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred4))
```

Bikin Confussion Matrix dari RFCS

```python
heatmap = confusion_matrix(ytest, ypred4)

plt.figure(figsize = (15,15))
sns.heatmap(heatmap, cmap='rocket', annot = True, fmt = 'g')
```
