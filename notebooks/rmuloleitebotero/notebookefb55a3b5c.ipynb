```python
#instanciando a base de traino

treino = pd.read_csv('/kaggle/input/airline-passenger-satisfaction/train.csv')
```

```python
#instanciando a base de teste

teste = pd.read_csv('/kaggle/input/airline-passenger-satisfaction/test.csv')
```

```python
#Analise Exploratória
#Verificando nulos no campo 'Arrival Delay in Minutes'


treino[treino['Arrival Delay in Minutes'].isna()]
```

```python
#Analise Exploratória
#Verificando nulos no campo 'Arrival Depature in Minutes'
treino[treino['Departure Delay in Minutes'].isna()]
```

```python
#Tratamento dos dados

#Substituindo na por 0
treino = treino.fillna(0)
teste = teste.fillna(0)
#transformando em inteiro 'Arrival Delay in Minutes'
teste['Arrival Delay in Minutes'] = teste['Arrival Delay in Minutes'].astype(int)
```

```python
#Analise Exploratória

treino.shape, teste.shape
```

```python
#Analise Exploratória

treino.info()
```

```python
#Analise Exploratória

teste.info()
```

```python
#Tratamento dos dados

#transformando "satisfied" em 1 e "neutral or dissatisfied" em 0

mapeamento = {'satisfied':1, 'neutral or dissatisfied':0}
treino['satisfaction'] = treino['satisfaction'].replace(mapeamento).astype(int)
teste['satisfaction'] = teste['satisfaction'].replace(mapeamento).astype(int)
```

```python
#Analise Exploratória

treino['satisfaction']
```

```python
#Retirando variáveis não númericas e a variável targert:  'satisfaction'

nao_usadas = ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'Unnamed: 0', 'id', 'satisfaction' ]
```

```python
#Instanciando variável com as variáveis(campos) que serão usadas

usadas = [c for c in treino.columns if c not in nao_usadas]
```

```python
#Analise Exploratória

usadas
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs=1, n_estimators=200, random_state=77)
```

```python
rf.fit(treino[usadas], treino['satisfaction'])
```

```python
teste['satisfaction'] = rf.predict(teste[usadas]).astype(int)
```

```python
#porcentagem de passageiros 0 = não satisfeitos e 1 = para satisfeitos na base de treino

treino['satisfaction'].value_counts(normalize=True)
```

```python
#porcentagem de passageiros 0 = não satisfeitos e 1 = para satisfeitos na base de teste

teste['satisfaction'].value_counts(normalize=True)
```

```python
fig=plt.figure(figsize=(10,10))
pd.Series(rf.feature_importances_,index=usadas).sort_values().plot.barh()

#variáveis com maior importância para determinar a satisfação do cliente
```

```python
# Conclusão
# A base de treino apresentou 56,6% de clientes não satisfeitos e 43,3% de clientes satisfeitos
# A base de teste apresentou 57,7% de clientes não satisfeitos e 42,2% de clientes satisfeitos
# A diferença na variável resposta foi de aproximadamente 1%, o que nos mostras que o modelo foi eficiente
```
