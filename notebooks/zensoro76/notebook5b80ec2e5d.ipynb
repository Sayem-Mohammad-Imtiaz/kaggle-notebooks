```python
df = pd.read_csv('/kaggle/input/weather-dataset-rattle-package/weatherAUS.csv')
df
```

```python
df.Location.unique()
df = df.loc[df.Location == 'Portland']
df = df.sort_values('Date')
df = df.head(400)
```

**Question 2:**

```python
df.plot(x = 'Date', y = 'Rainfall', title = 'FirstLineGraph')
```

**Question 3/4**

```python
df['RainRolling5'] = df.Rainfall.rolling(5).mean()
df.plot(x = 'Date', y = 'RainRolling5', title = 'FirstLineGraph')
```

```python
plt.plot(df.Date, df.Rainfall, df.Date, df.RainRolling5)
plt.show()
```

```python
df['RainRolling10'] = df.Rainfall.rolling(10).mean()
df.plot(x = 'Date', y = 'RainRolling10', title = 'FirstLineGraph')
```

```python
plt.plot(df.Date, df.Rainfall, df.Date, df.RainRolling10)
plt.show()
```

```python
df['RainRolling3'] = df.Rainfall.rolling(3).mean()
df.plot(x = 'Date', y = 'RainRolling3', title = 'FirstLineGraph')
```

```python
plt.plot(df.Date, df.Rainfall, df.Date, df.RainRolling3)
plt.show()
```

```python
df['RainRolling20'] = df.Rainfall.rolling(20).mean()
df.plot(x = 'Date', y = 'RainRolling20', title = 'FirstLineGraph')
```

```python
plt.plot(df.Date, df.Rainfall, df.Date, df.RainRolling20)
plt.show()
```

**Question 5**

I believe that the rolling mean with a window of 10 shows the data in the most accurat eand useful way. While the data is still all over the place, this is to be predicted with rain. The rolling mean helped show that the rain is starting to slow down. This is seen by the rolling mean beginning to decrease. This would have been difficult but possible to see from the original graph but the rolling mean made it much more clear. 

**Question 6**

A smaller mean window is going to look more similar to the original graph and may include points that do not fit well with the graph. A larger window is likely to remove those datapoints, but it will average a larger amount of data. This may lead to smaller trends being lost in the larger window.

**Quesiton 7**

A window of 1 would look exactly like the orignial data.
A window of the length of the dataframe would be the average value of the y value.
If the window was half the data, it would show a much softer curve in the data.
