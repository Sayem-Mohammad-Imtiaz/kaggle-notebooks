```python
df4 = pd.get_dummies(df3,columns =['gender','hypertension','heart_disease','ever_married','smoking_status'],drop_first=True)
df4.head()
```

```python
columns ={"gender_Male":"Is_Male","hypertension_1":"Is_hypertens","heart_disease_1":"Has_heart_disease","ever_married_Yes":"Is_Married","smoking_status_formerly smoked":"quit_smoking",
          "smoking_status_never smoked":"never_smoked","smoking_status_smokes":"smoker"}
```

```python
df5 = df4.rename(columns = columns)
df5.head()
```

```python
X = df5[['age','avg_glucose_level','bmi','Is_Male','Is_hypertens','Has_heart_disease','quit_smoking','never_smoked','smoker']]
X.head()
```

```python
y = df5.stroke
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state =999)   #we use only 30 percent for testing and random state is used for reproducibility
                                                                                        #(get the same split on all computers)
```

```python
from sklearn.linear_model import RidgeClassifier  #to penalize the slope by increasing bias and reducing variance for consistent results(the bias-variance tradeoff)
```

```python
clf = RidgeClassifier()
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
y_test[:10]
```

```python
y_pred[:10]
```

```python
metrics.accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
cm=metrics.confusion_matrix(y_test,y_pred)
```

```python
sns.heatmap(cm/np.sum(cm),annot=True,fmt='.2%',cmap="Greens")
```
