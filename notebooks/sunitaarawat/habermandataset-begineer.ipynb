```python
#importing all the libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
haberman = pd.read_csv("/kaggle/input/habermans-survival-data-set/haberman.csv", header = None, names=['age', 'year', 'nodes', 'status'])
```

```python
haberman.head(3)
```

```python
#(Q) What are the column names in our dataset?
print (haberman.columns)
```

```python
#How many data points for each class are present? 
haberman["status"].value_counts()
```

Observations: -- The value_counts() function tells how many data points for each class are present. Here, it tells how many patients survived and how many did not survive. -- Out of 306 patients, 225 patients survived and 81 did not. -- The dataset is imbalanced.

```python
#brief info about the dataset
print(haberman.info())
```

### Observations:
---- There are no missing values in this data set. <br/>
---- All the columns are of the integer data type.<br/>
---- The datatype of the status is an integer, it has to be converted to a categorical datatype <br/>
---- In the status column, the value 1 can be mapped to ‘yes’ which means the patient has survived 5 years or longer. And the value 2 can be mapped to ‘no’ which means the patient died within 5 years.<br/>

```python
#2d scatter plot
#******************************************************************************
haberman.plot(kind='scatter',x='nodes',y='age')
plt.title("2-d scatter plot /w nodes and age")
plt.show()
#this plot doesn't give any information about the classes
```

```python
#2d scatter plot with different colors 
#*************************************************
sns.set_style("darkgrid");
sns.FacetGrid(haberman, hue="status", height=5).map(plt.scatter, "nodes", "age").add_legend();
plt.title('2-D scatter plot Age Vs Year')
plt.show();
```

Observation of above,classes are overlap eachother. The plot gives us a rough idea that both the classes are mixed and and not easily distinguishable

```python
#this pair plot gives class dependancies w.r.to each variable/feature
plt.close();
sns.set_style("darkgrid");
sns.pairplot(haberman, hue="status", height=3, vars = ["nodes","age","year"]);
plt.suptitle('Pair Plot for each feature')
plt.show()
```

Observation : classes are not easily separable. Every variable has the mixed representation of both the classes. That's why we can't draw a linearly separable line from any of them(Plots).

```python

#Histogram, pdf, cdf
#1-D scatter plot of AGE
#***************************************
haber_one = haberman.loc[haberman["status"] == 1];
haber_two = haberman.loc[haberman["status"] == 2];

plt.plot(haber_one["age"], np.zeros_like(haber_one['age']), 'o')
plt.plot(haber_two["age"], np.zeros_like(haber_two['age']), 'r--')

plt.legend('12')
plt.title('1-D scatter plot for age')
plt.show()
```

```python
# Histogram and PDF of AGE for both classes
#****************************************************************
sns.FacetGrid(haberman, hue="status", height=5) .map(sns.distplot, "age").add_legend();
plt.title('Histogram and PDF of AGE')
plt.ylabel('count')
plt.show()
```

Observations :<br/> - the data for both the class range between age 30 to 85.
            both class have mean approximatly equal.<br/>
               - Major overlapping is observed, which tells us that survival chances are irrespective of a person’s age.

```python
# Histogram and PDF of Op_Year for both classes
#************************************************
sns.FacetGrid(haberman, hue="status", height=5).map(sns.distplot, "year").add_legend();
plt.title('Histogram and PDF of Year')
plt.ylabel('count')
plt.show();
```

Observations :<br/>- With respect to "Year" both the classes are overlaped and this feature not readily differentiate them.<br/>
-- it can be observed that in the years 1960 and 1965 there were more unsuccessful operations

```python
sns.FacetGrid(haberman, hue="status", height=5).map(sns.distplot, "nodes").add_legend();
plt.title('Histogram and PDF of Nodes')
plt.ylabel('count')
plt.show();
```

observations :<br/>  Patients with no nodes or 1 node are more likely to <br/>survive. There are very few chances of surviving if there are 25 or more nodes.

```python

#CDF 
# CDF plot of Age for class 1
#***************************************************************************************************
counts, bin_edges = np.histogram(haber_one['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
#plt.xlabel("age")
#plt.ylabel("Probability")
#plt.title("CDF plot of Age for class 1")



print("********************************************************")
counts2, bin_edges2 = np.histogram(haber_two['nodes'], bins=10, 
                                 density = True)
pdf2 = counts2/(sum(counts2))
print(pdf2);
print(bin_edges2);
cdf2 = np.cumsum(pdf2)
plt.plot(bin_edges2[1:],pdf2);
plt.plot(bin_edges2[1:], cdf2)

#plt.xlabel("age")
#plt.ylabel("Probability")
plt.legend('12')
plt.show();
```

```python
#Box-plot can be visualized as a PDF on the side-ways.
sns.boxplot(x='status',y='age',data=haberman)
plt.show()
sns.boxplot(x='status',y='year',data=haberman)
plt.show()
sns.boxplot(x='status',y='nodes',data=haberman)
plt.show()
```

```python
sns.violinplot(x="status",y="age",data = haberman,height = 10)
plt.show()
sns.violinplot(x="status",y="year",data = haberman,height = 10)
plt.show()
sns.violinplot(x="status",y="nodes",data = haberman,height = 10)
plt.show()
```

Observations:

-- Patients with more than 1 nodes are not likely to survive. More the number of nodes, lesser the survival chances.
-- There were comparatively more people who got operated in the year 1965 did not survive for more than 5 years.
-- There were comparatively more people in the age group 45 to 65 who did not survive. Patient age alone is not an important parameter in determining the survival of a patient.
