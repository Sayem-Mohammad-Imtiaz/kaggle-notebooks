```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
```

```python
data = pd.read_csv("/kaggle/input/gender-classification/Transformed Data Set - Sheet1.csv")
data.info()
```

```python
data.rename(columns={'Favorite Color' :'FavoriteColor', 'Favorite Music Genre':'FavoriteMusicGenre', 
                          'Favorite Beverage':'FavoriteBeverage', 'Favorite Soft Drink':'FavoriteSoftDrink'}, inplace=True)
```

```python
#Fetch features of type Object
objFeatures = data.select_dtypes(include="object").columns

#Iterate a loop for features of type object
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for feat in objFeatures:
    data[feat] = le.fit_transform(data[feat].astype(str))
    
data.info()
```

```python
data
```

```python
accuracies = {}
X = data.drop(['Gender'], axis = 1)
Y = data.Gender
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(), annot = True)
```

# **LABA 2B**

**Tanh**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

mlp_tanh = MLPClassifier(hidden_layer_sizes=25, activation='tanh', max_iter = 800)
mlp_tanh.fit(X_train, Y_train)
```

```python
probability_tanh = mlp_tanh.predict_proba(X_test)
print(probability_tanh)
```

```python
Y_pred = mlp_tanh.predict(X_test);
```

```python
mlp_tanh_score = mlp_tanh.score(X_test, Y_test)*100

print('Train accuracy: ','{:.2f}'.format(mlp_tanh.score(X_train, Y_train)*100),"%")
print('Test accuracy: ','{:.2f}'.format( mlp_tanh_score),"%")

accuracies['tahn'] = mlp_tanh_score
```

```python
print(' layers: ',mlp_tanh.n_layers_, '\n','iterations: ',mlp_tanh.n_iter_)
```

```python
cm_tanh = confusion_matrix(Y_test,Y_pred)
plt.title("Confusion Matrix")
sns.heatmap(cm_tanh,annot=True,cmap="Greens",fmt="d",cbar=False, annot_kws={"size": 10})
```

**Relu**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

mlp_relu = MLPClassifier(hidden_layer_sizes=25, activation='relu', max_iter = 800)
mlp_relu.fit(X_train, Y_train)
```

```python
probability_relu = mlp_relu.predict_proba(X_test)
print(probability_relu)
```

```python
Y_pred = mlp_relu.predict(X_test);
```

```python
mlp_relu_score = mlp_relu.score(X_test, Y_test)*100

print('Train accuracy: ','{:.2f}'.format(mlp_relu.score(X_train, Y_train)*100),"%")
print('Test accuracy: ','{:.2f}'.format( mlp_relu_score),"%")

accuracies['relu'] = mlp_relu_score
```

```python
print(' layers: ',mlp_relu.n_layers_, '\n','iterations: ',mlp_relu.n_iter_)
```

```python
cm_relu = confusion_matrix(Y_test,Y_pred)
plt.title("Confusion Matrix")
sns.heatmap(cm_relu,annot=True,cmap="Greens",fmt="d",cbar=False, annot_kws={"size": 10})
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(9,7))
plt.ylabel("Accuracy")
plt.xlabel("Algorithms")
sns.barplot(x=list(accuracies.keys()), y=list(accuracies.values()), palette=["blue", "green", "orange"])
plt.show()
```

# **LABA3**

**Decision Tree**

```python
accuracies1 = {}

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5)

dt_classifier = DecisionTreeClassifier(max_depth = 7, splitter = 'random', min_samples_split = 2)
dt_classifier.fit(X_train, Y_train)

Y_pred = dt_classifier.predict(X_test)
```

```python
dt_classifier_score = dt_classifier.score(X_test, Y_test)*100

print('Train accuracy: ','{:.2f}'.format(dt_classifier.score(X_train, Y_train)*100),"%")
print('Test accuracy: ','{:.2f}'.format( dt_classifier_score),"%")

accuracies1['decision tree'] = dt_classifier_score
```

```python
cm_dt_classifier = confusion_matrix(Y_test,Y_pred)
plt.title("Confusion Matrix")
sns.heatmap(cm_dt_classifier,annot=True,cmap="Greens",fmt="d",cbar=False, annot_kws={"size": 10})
```

```python
import graphviz 
fig = plt.figure(figsize=(20,25))
dot_data = tree.export_graphviz(dt_classifier, out_file=None,  
                     filled=True, rounded=True,  
                      special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

**RandomForest**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5)

rf_classifier = RandomForestClassifier(n_estimators = 20, criterion = 'gini')
rf_classifier.fit(X_train, Y_train)

Y_pred = rf_classifier.predict(X_test)
```

```python
rf_classifier_score = rf_classifier.score(X_test, Y_test)*100

print('Train accuracy: ','{:.2f}'.format(rf_classifier.score(X_train, Y_train)*100),"%")
print('Test accuracy: ','{:.2f}'.format( rf_classifier_score),"%")

accuracies1['random forest'] = rf_classifier_score
```

```python
cm_rf_classifier = confusion_matrix(Y_test,Y_pred)
plt.title("Confusion Matrix")
sns.heatmap(cm_rf_classifier,annot=True,cmap="Greens",fmt="d",cbar=False, annot_kws={"size": 10})
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(9,7))
plt.ylabel("Accuracy")
plt.xlabel("Algorithms")
sns.barplot(x=list(accuracies1.keys()), y=list(accuracies1.values()), palette=["green", "orange"])
plt.show()
```
