```python
from matplotlib import pyplot as plt
import seaborn as sns
ipl = pd.read_csv('/kaggle/input/ipl-data-set/matches.csv')
ipl.head()
```

```python
ipl['player_of_match'].value_counts()
```

```python
ipl.info()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()), list(ipl['player_of_match'].value_counts()[0:5]),color = 'r')
plt.show()
```

```python
ipl['result'].value_counts()
```

```python
plt.bar(ipl['result'].value_counts().keys(),ipl['result'].value_counts(),color = 'g')
plt.show()
```

```python
batting_first = ipl[ipl['win_by_runs']!=0]
print(batting_first)
```

```python
batting_first.head()
```

```python

plt.figure(figsize =(10,7))
plt.hist(ipl['win_by_runs'], color = 'y')
plt.xlabel('runs')
plt.title('victory')
plt.show()
```

```python
ipl['winner'].value_counts()
```

```python
plt.bar(ipl['winner'].value_counts()[0:3].keys(),ipl['winner'].value_counts()[0:3])
plt.show()
```

```python
plt.figure(figsize = (20,15))
plt.pie(list(batting_first['winner'].value_counts()),labels = list(batting_first['winner'].value_counts().keys()),autopct = '%0.1f%%')
plt.show()
```
