```python
df = pd.read_csv("/kaggle/input/pima-indians-diabetes-database/diabetes.csv")
df.head()
```

```python
x = df[['Glucose', 'BloodPressure']]
y = df['Outcome']
x, y
```

```python
# sigmoid function
def sigmoid(input):
    output = 1/ (1 + np.exp(-input))
    return output
```

```python
#sigmoid derivative function
def sigmoid_derivative(input):
    return sigmoid(input) * (1.0 - sigmoid(input))
```

```python
def train_network(features, label, weights, bias, learning_rate, epochs):
    for epoch in range(epochs):
        #dot product
        dot_prod = np.dot(features, weights) + bias
        #using sigmoid
        preds = sigmoid(dot_prod)
        #Calculating the error
        errors = preds - label
#         print(preds[:5])
#         print(label[:5])
#         print(errors[:5])
        #sigmoid derivative
        derivative_cost_funct = errors
        derivative_preds = sigmoid_derivative(preds)
        derivative_product = derivative_cost_funct * derivative_preds
        features_T = features.T
        #update the weights
        weights = weights - np.dot(features_T, derivative_product) * learning_rate
        loss = errors.sum()
        print(loss)
    for i in derivative_product:
        bias = bias - i * learning_rate
    return  weights, bias, preds
```

```python
np.random.seed(10)
features = x
label = y.values.reshape(y.shape[0], 1)
weights = np.random.rand(2, 1)
bias = np.random.rand(1)
learning_rate = 0.0000004
epochs = 300
```

```python
weights_final, bias_final, preds = train_network(features, label, weights, bias, learning_rate, epochs)
```
