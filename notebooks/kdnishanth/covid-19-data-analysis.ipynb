# What is Novel Coronavirus?
The novel coronavirus (provisionally named 2019-nCoV) is a contagious virus that causes respiratory infection. It has been identified as the causative agent of the ongoing 2019â€“20 Wuhan coronavirus outbreak.

As many early cases were linked to a large seafood and animal market, the virus is thought to have a zoonotic origin, but this has not been confirmed. Comparisons of the genetic sequences of this virus and other virus samples have shown similarities to SARS-CoV (79.5%) and bat coronaviruses (96%), which makes an ultimate origin in bats likely.

The first known human infection occurred in December 8, 2019. An outbreak of 2019-nCoV was first detected in Wuhan, China, in mid-December 2019.The virus subsequently spread to all other provinces of China and to more than twenty other countries in Asia, Europe, North America, and Oceania. Human-to-human spread of the virus has been confirmed in China, Germany, Thailand, Taiwan, Japan, and the United States.

As of 1 February 2020, there were 12,024 confirmed cases of infection, of which 11,860 were within mainland China. Cases outside China, to date, were people who have either travelled from Wuhan, or were in direct contact with someone who travelled from the area. The number of deaths was 259 as of 1 February 2020.

Source: https://en.wikipedia.org/wiki/Novel_coronavirus_(2019-nCoV)

```python
Input_Data = pd.read_csv('../input/covid19-in-india/covid_19_india.csv')
Input_Data.head(5)
```

Removing the Duplicate State name's 

```python
_list = []
for i in Input_Data["State/UnionTerritory"]:
    _list.append(i)
mylist = dict.fromkeys(_list)
District = []
for x in mylist.keys():
    District.append(x)
print(District)
```

Finding the Affected,Recurved and Died People

```python
_cured = [];confirmed = [];_death = [];_District = []
for i in range(len(District)):
    _sum = 0;death = 0;cured = 0
    _District.append(District[i])
    for j in range(len(Input_Data["State/UnionTerritory"])):
        if District[i] == Input_Data["State/UnionTerritory"][j]:
            cured = cured + Input_Data["Cured"][j]
            _sum = _sum + Input_Data["Confirmed"][j]
            death = death + Input_Data["Deaths"][j]
    _cured.append(cured)
    confirmed.append(_sum)        
    _death.append(death) 
```

## displaying the Confirmed Case's

```python
from matplotlib import style
fig = plt.figure(figsize=(10,10))
ax1 = plt.subplot2grid((1,1), (0,0))
plt.plot( District, confirmed,  color='blue', linestyle='dashed', linewidth = 1, 
             marker='o', markerfacecolor='orange', markersize=15, label='Count')
for label in ax1.xaxis.get_ticklabels():
        label.set_rotation(85)  
style.use('ggplot')
plt.legend() 
plt.xlabel("State's", size = 20)
plt.ylabel("Number of Person's Affected\ndue to COVID-19\n", size = 15)
plt.show()
```

## Displaying the Recovered Case's 

```python
from matplotlib import style
fig = plt.figure(figsize=(10,10))
ax1 = plt.subplot2grid((1,1), (0,0))
plt.plot( District, _cured, color='k', linestyle='dashed', linewidth = 1, 
             marker='o', markerfacecolor='blue', markersize=15, label='Recurvered') 
for label in ax1.xaxis.get_ticklabels():
        label.set_rotation(85)  
style.use('ggplot')
plt.legend() 
plt.xlabel("State's", size = 20)
plt.ylabel("Number of students recovered\nfrom COVID-19\n", size = 20)
plt.show()
```

## Displaying the Death Case's 

```python
from matplotlib import style
fig = plt.figure(figsize=(10,10))
ax1 = plt.subplot2grid((1,1), (0,0))
plt.plot(District, _death, color='green', linestyle='dashed', linewidth = 1, 
             marker='o', markerfacecolor='red', markersize=15, label='Death') 
for label in ax1.xaxis.get_ticklabels():
        label.set_rotation(85)  
style.use('ggplot')
plt.legend() 
plt.xlabel("State's", size = 20)
plt.ylabel("Number of Person's People Died\ndue to COVID-19\n", size = 20)
plt.show()
```
