```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('../input/world-happiness-report-2021/world-happiness-report-2021.csv')
```

```python
data
```

SELECTING IMPORTANT VARIABLE AND DROPING THE REMAINING AS MENTION IN DATA DESCRIPTION

```python
data.columns
```

```python
data1=data[['Country name','Regional indicator','Ladder score','Logged GDP per capita',
         'Social support','Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption',
         'Ladder score in Dystopia']]
```

```python
data1.shape
```

```python
data1.isnull().sum() #here no missing value is found
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data=data1,x='Regional indicator',y='Ladder score')
plt.xticks(rotation=45)
```

### from above graph we can see that north america and ANZ has the highest ladder score/happiness score and south asia has the lowest score

## lets find the correlation of happiness score/Ladder score with other variable

```python
data1.corr()[['Ladder score']].sort_values(by='Ladder score',ascending=False)
```

## as we can see that GDP,HEATH,Social Support,Freedom to make life choices are positively correlated to happiness score                                                                                                                 but nobality(generosity)is negatively correlated with happiness score ,its very strange                                                                                                              


```python
top_10_happiest=data1.sort_values(by='Ladder score',ascending=False).head(10)
top_10_unhappiest=data1.sort_values(by='Ladder score',ascending=False).tail(10)
```

```python
plt.figure(figsize=(20,6))
sns.barplot(data=data1,x=top_10_happiest['Country name'],y=top_10_happiest['Ladder score'])
```

## as we can see from above graph that Finland is happiest country in this world 

## 

```python
plt.figure(figsize=(20,6))
sns.barplot(data=data1,x=top_10_unhappiest['Country name'],y=top_10_unhappiest['Ladder score'])
```

## 

## as we can see that Afaganistan is the unhappiest country in the world as it has lowest ladder score
