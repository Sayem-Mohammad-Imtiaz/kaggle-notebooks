 
## Number of Suicides over the time

```python
df.plot(kind="scatter",x="year",y="suicides_no",alpha=0.5,color="red",figsize=(10,10),grid=True,label="Suicides No - Year")
plt.xlabel("YEAR")
plt.ylabel("SUICIDES NO")
plt.legend()
plt.title("Suicides in Over Time")
plt.show()
```

## Country which have highest Suicides rate

```python
df.plot(kind='line',x='country',y='suicides_no',alpha=0.8,color='red',
        figsize=(10,10),
        grid=True,label='Suicides No - Country')
plt.xlabel('country')
plt.ylabel('sucides no')
plt.legend()
plt.title('Suicide number based on country')
plt.show()
```

## Suicides based on Sex

```python
df.plot(kind='scatter',x='sex',y='suicides_no',alpha=0.8,color='red',
        figsize=(10,10),
        grid=True,label='Suicides No - Sex')
plt.xlabel('sex')
plt.ylabel('suicides no')
plt.legend()
plt.title('Suicides number based Sex')
plt.show()
```

## Age which have highest rate of suicide

```python
df.plot(kind="line",x="age",y="suicides_no",label="SUICIDES NO",grid=True,alpha=.7,color='red',figsize=(10,10))
plt.legend()
plt.xlabel("AGE")
plt.ylabel("SUICIDES NO")
plt.title("SUICIDES-AGE CONNECTION")
plt.show()
```
