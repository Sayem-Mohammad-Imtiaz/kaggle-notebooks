# Middle East Countries

Classifications of the countries in the middle east vary from one source to another. However, I only took in account 10 countries to analyze (Bahrain, United Arab Emirates, Saudi Arabia, Cyprus, Kuwait, Iraq, Iran, Lebanon, Jordan, Yemen). It also maight be useful to indicate that 8 countries of the 10 belong to the Arab world (Bahrain, United Arab Emirates, Sadui Arabia, Kuwait, Iraq, Jordan, Yemen)  

```python
Middle_East = ["Bahrain","United Arab Emirates","Saudi Arabia","Cyprus","Kuwait","Iraq","Iran","Lebanon",
               "Jordan","Yemen"]
```

# Ladder Score by country ranked

```python
life = World_Happiness2021Report.groupby("Country name")["Ladder score"].mean().reset_index()
life["Index"] = [i for i in range(len(life))] 
life = life[life["Country name"].isin(Middle_East)] 
life.columns = ['Country name', 'Ladder score', 'Rank'] 
life = life.loc[life["Ladder score"].nlargest(10).index.tolist(),:] 
life["Rank"] = [i for i in range(1,len(life["Rank"])+1)] 

ff.create_table(life)
```

```python
MidEast = World_Happiness[World_Happiness["Country name"].isin(Middle_East)] 
plt.figure(figsize=(15,10))
px.line(MidEast,x="year",y="Life Ladder",color="Country name",template="plotly_dark")
```

# Logged GDP per capita by country ranked

```python
Logged_GDP = World_Happiness2021Report.groupby("Country name")["Logged GDP per capita"].mean().reset_index()
Logged_GDP["Index"] = [i for i in range(len(Logged_GDP))] 
Logged_GDP = Logged_GDP[Logged_GDP["Country name"].isin(Middle_East)] 
Logged_GDP.columns = ['Country name', 'Logged GDP per capita', 'Rank']
Logged_GDP = Logged_GDP.loc[Logged_GDP["Logged GDP per capita"].nlargest(10).index.tolist(),:] 
Logged_GDP["Rank"] = [i for i in range(1,len(Logged_GDP["Rank"])+1)] 

ff.create_table(Logged_GDP)
```

```python
MidEast = World_Happiness[World_Happiness["Country name"].isin(Middle_East)] 
plt.figure(figsize=(15,10))
px.line(MidEast,x="year",y="Log GDP per capita",color="Country name",template="plotly_dark")
```
