# **From Amazons Top50 Bestselling Books, which author writes best and cheapest books?**

Importing neccesary libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Loading data

```python
books = pd.read_csv('/kaggle/input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
books.head()
```

How many unique authors are there?

```python
books['Author'].nunique()
```

```python
books['Author'].value_counts()
```

Creating new DataFrame of Authors and applying number of their books

```python
authors = pd.DataFrame(books['Author'].unique(),columns=['Name'])
authors
```

Number of books for each author

```python
cnt = books['Author'].value_counts()
authors['NbBooks'] = authors['Name'].apply(lambda x: cnt[x])
authors.sort_values('NbBooks',ascending=False)
```

Means of their prices and raitings

```python
mop = books.groupby('Author').mean()['Price']
authors['MoPrice'] = authors['Name'].apply(lambda x: mop[x])

mour = books.groupby('Author').mean()['User Rating']
authors['MoUserRating'] = authors['Name'].apply(lambda x: mour[x])

authors
```

Checking who has more than 1 book in Top50

```python
authors['MultiBooks'] = authors['NbBooks'].apply(lambda x: x>1)
authors
```

**Plot ranking x Price**

```python
sns.jointplot(data=authors,x='MoUserRating',y='MoPrice',kind='hex',color='red')
```

Dividing means of users rating and price gives factor, which rise with better rating and lower price

```python
authors['RatingPriceFactor'] = authors['MoUserRating']/authors['MoPrice']
```

```python
authors=authors[authors['RatingPriceFactor']<np.inf]
authors.sort_values('RatingPriceFactor',ascending=False)
```

Cheapest && Best rated authors with only 1 book in Top50:

```python
authors[authors['MultiBooks']==False].sort_values('RatingPriceFactor',ascending=False)['Name'][:10]
```

Cheapest && Best rated authors with more than 1 book in Top50:

```python
authors[authors['MultiBooks']].sort_values('RatingPriceFactor',ascending=False)['Name'][:10]
```

Most expesnsive && worst rated authors

```python
authors.sort_values('RatingPriceFactor')['Name'][:10]
```
