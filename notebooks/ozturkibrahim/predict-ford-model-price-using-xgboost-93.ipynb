* Let's revise the data as the engine size cannot be 0.0.

```python
round(pd.pivot_table(data = df, columns = "engineSize", values = "price")).T
```

**Let's move on to the machine learning part
**

```python
from xgboost import XGBRegressor
from sklearn.model_selection import train_test_split, GridSearchCV,cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import model_selection
```

```python
#import openpyxl
dflbl = pd.read_excel(r"dflbl.xlsx",engine='openpyxl')
```

```python
X = dflbl.drop(["price"], axis = 1) # We separated dependent and independent variables price dependent variable
y = dflbl["price"] #the dependent variable
```

```python
# Separation of training and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 144)
```

```python
# we find the most suitable parameters
params = {"colsample_bytree":[0.4,0.5,0.6],
         "learning_rate":[0.01,0.02,0.09],
         "max_depth":[2,3,4,5,6],
         "n_estimators":[100,200,500,2000]}

xgb = XGBRegressor()

grid = GridSearchCV(xgb, params, cv = 10, n_jobs = -1, verbose = 2)

grid.fit(X_train, y_train)

grid.best_params_
```

```python
# we enter the most appropriate parameters
xgb1 = XGBRegressor(colsample_bytree = 0.5, learning_rate = 0.02, max_depth = 6, n_estimators = 2000)
```

```python
# We train our model
model_xgb = xgb1.fit(X_train, y_train)
```

```python
# We calculated the model's score. Value between 0-1 returns us
model_xgb.score(X_test, y_test)
```

```python
model_xgb.score(X_train, y_train)
```

```python
#validated (we find our verified error rate)
np.sqrt(-1*(cross_val_score(model_xgb, X_test, y_test, cv=10, scoring='neg_mean_squared_error'))).mean()
```
