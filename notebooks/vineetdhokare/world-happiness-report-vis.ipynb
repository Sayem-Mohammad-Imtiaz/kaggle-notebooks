```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from scipy import stats

# for min_max scaling
from mlxtend.preprocessing import minmax_scaling
print("Setup Complete")
```

```python
report_2015 = pd.read_csv('../input/world-happiness/2015.csv')
report_2015.head()
```

```python
report_2015.columns
```

```python

sns.regplot(x = report_2015['Happiness Score'], y = report_2015['Economy (GDP per Capita)'])
```

The above graph clearly shows that countries having higher gpd tend to have higher happiness score

```python
sns.pairplot(report_2015, vars=['Happiness Score',
       'Standard Error', 'Economy (GDP per Capita)', 'Family',
       'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)',
       'Generosity', 'Dystopia Residual'])
```

```python
#taking columns containing numerical values
numerical_report_2015 = report_2015.loc[:,'Happiness Rank':'Dystopia Residual']
numerical_columns = list(numerical_report_2015.columns[1:-1])
scaled_report = minmax_scaling(numerical_report_2015, columns = numerical_columns)
scaled_report
```

```python
plt.figure(figsize = (12,6))
#corelation heatmap between all the variables (happiness rank is obviously not included)
sns.heatmap(data = report_2015.loc[:,'Happiness Score':'Dystopia Residual'].corr(), cmap='coolwarm', annot = True)
```

It looks like a Countries Happieness Score is greatly dependent on 3 factors:
'GDP', 'Life Expectancy' and 'Family'
