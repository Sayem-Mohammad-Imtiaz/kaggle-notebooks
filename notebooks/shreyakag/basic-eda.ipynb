# Univariate analysis
*  It's major purpose is to describe; It takes data, summarizes that data and finds patterns in the data.

# Histograms and Probability Density Function 
* Hisograms are basically substitute for graph for univariate analysis without overlapping of datapoints.
* PDF is smoothned form of histogram.Height of histogram equals to number of data points[density]. 

```python
df.hist()
plt.show()
```

```python
sns.FacetGrid(df,hue="Survival status",height = 5)\
 .map(sns.distplot,"Age")\
 . add_legend();
plt.show()
```

# observations
*  Survival status cant be concluded based on age because of overlapping of data points.
* Loose conclusions can be made like 30-40 aged are likely to survive,40-60 may not survive,60-75 cant be said.
* So Age is not correct feature to anaylze survival status.
 

```python
sns.FacetGrid(df,hue="Survival status",height = 5)\
 .map(sns.distplot,"Year operation")\
 . add_legend();
plt.show()
```

# Observations
* Clearly year of operation is not a feature dependent of status of survival because of the overlaps.
* Nonetheless 62.5 and 67.5 years has some success.

```python
sns.FacetGrid(df,hue="Survival status",height = 5)\
 .map(sns.distplot,"Axillary nodes detected")\
 . add_legend();
plt.show()
```

# Observations
* There are less overlaps in this case.
* Patients with 0 or 1 node is more likely to survive.

# Box plots 
The box extends from the lower to upper quartile values of the data, with a line at the median. The whiskers extend from the box to show the range of the data. Outlier points are those past the end of the whiskers.

```python
sns.boxplot(x='Survival status',y='Age',data=df)
plt.show()
sns.boxplot(x='Survival status',y='Year operation',data=df)
plt.show()
sns.boxplot(x='Survival status',y='Axillary nodes detected',data=df)
plt.show()
```

# observations
* Age feature is overlapping ,year 65-66 survived,0/1 nodes survived.


# violin plot
* It offers best of both the worlds[histogram and boxplots]
* Number of data points as well as percentiles can be seen

```python
sns.violinplot(x='Survival status',y='Age',data=df)
plt.show()
sns.violinplot(x='Survival status',y='Year operation',data=df)
plt.show()
sns.violinplot(x='Survival status',y='Axillary nodes detected',data=df)
plt.show()
```

# observations
* Node 0 survived.
* Age is not a factor in terms of survival.

# Bivariate Analysis
* Scatter plots 
* pair plots

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue="Survival status",height = 6)\
 .map(plt.scatter,"Age","Year operation")\
 . add_legend();
plt.show()
```

# observations
* Too  much overlapping. so nothing can be determind.

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue="Survival status",height = 6)\
 .map(plt.scatter,"Age","Axillary nodes detected")\
 . add_legend();
plt.show()
```

# observations
* Only 3-4 people with 20+ nodes survived.
* Most survived people are ranging from nodes 0-10.
* Seems that age is not a major factor in survival.

```python
sns.set_style("whitegrid")
sns.pairplot(df,hue="Survival status",height = 6)
plt.show()
```

# observations
* Plot between node and age is good without much overlaps along with node and year.

# Multivariate analysis
* Its also called contour probality density plot.
* The colors are called contours.

```python
sns.jointplot(x = "Year operation", y = "Age", data = df, kind = "kde")
plt.show()
sns.jointplot(x = "Year operation", y = "Axillary nodes detected", data = df, kind = "kde")
plt.show()
```

# observations
* Irrespective of year node 0 patients seems to survive more.
* In year 60-66 and operations are done  on 50-60 aged  people

# Final obeservations
1. Node 0 and 1 survived irrespective of other features.
2. Age below 40 have more chances of survival.
3. So if a new data point is given which satisfies these criteria they are likely to survive.Otherwise since the data is very overlapping it can be hard to say.
