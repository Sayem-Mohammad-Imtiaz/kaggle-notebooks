```python
data1['Price']=data1['Price'].str.replace(',', '')
data1['Price']=data1['Price'].str.replace('â‚¹', '')
data1['Price']=data1['Price'].apply(lambda x:0 if x=='Free'else x)
data1['Price']=data1['Price'].astype("float64")
```

```python
#create time span column
from datetime import date
today = date.today()
today=pd.to_datetime(today)
data1['Date'] = pd.to_datetime(data1['Date'])
data1['Duration'] = today-data1['Date']
data1['Duration']=data1['Duration'].dt.days
data1['Duration'].min()
```

```python
data1['Durationbin']=pd.cut(data1['Duration'],[-200,0,365,730,1095,1460,1825,3850],labels=["Not Released","1 Year","2 Years","3 Years",
                                                                                          "4 Years","5 Years","More Than 5 Years"])
data1['Price'].value_counts()
```
