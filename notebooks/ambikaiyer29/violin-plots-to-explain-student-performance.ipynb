```python
student_data = pd.read_csv("../input/StudentsPerformance.csv")
```

```python
student_data.head()
```

#### Derive the average score of the students across all subjects.

```python
student_data['overall_average'] = round((student_data['math score']+student_data['reading score']+student_data['writing score'])/3.0,2)
```

> ** Let's see the effect of lunch on the average score of the student **

```python
# sns.distplot(student_data.groupby(['gender']).reset_index().overall_average)
sns.violinplot(x="gender", y="overall_average", hue="lunch",split=True,data=student_data)
```

>** Clearly students with standard lunch perform better on an average compared to students with free/reduced lunch. This can indicate that the nutrition value of free/reduced lunches are not at par. **

```python
sns.violinplot(x="gender", y="overall_average", hue="parental level of education",data=student_data)
```

1. > **Looking at the violin plot above, it seems that the parents' education has little effect on the median score for females as compared to the male**

2. > **Male students with parents having a master's degree have the highest median and also the density is higher around the median**

```python
sns.catplot(x="gender", y="overall_average",
                hue="lunch", col="parental level of education",split=True,
               data=student_data, kind="violin");
```

> **An interesting thing to note regarding female students is that their overall_average score is in general either better than male students or atleast at par. However, for students with parents who have master's degree, female students seem to have a lower median value for the score.**

```python
sns.violinplot(x="test preparation course", y="overall_average", hue="gender",split=True,data=student_data)
```

> **This plot is expected. Students who have completed the test preparation course have a better score than those who haven't.**

```python
sns.violinplot(x="race/ethnicity", y="overall_average", hue="gender",split=True,data=student_data)
```

> **Again, the race/ethnicity seems to have a role to play in the student's scores. The medians are clearly varying with the race of the students.**
