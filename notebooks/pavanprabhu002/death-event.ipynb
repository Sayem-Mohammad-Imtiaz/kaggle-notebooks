```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Prediction problem.
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC


from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import f1_score
from sklearn.model_selection import train_test_split
```

# Data


```python
data = pd.read_csv('../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv')
```

```python
data

# We will try predicting the DEATH_EVENT.
```

# Target Feature


```python
y = data['DEATH_EVENT']
x = data.drop('DEATH_EVENT', axis=1)
```

# scaling the features into unit values

```python
scale = MinMaxScaler()
x = pd.DataFrame(scale.fit_transform(x))
x
```

# Training The Model

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.7)
```

# Logistic regression

```python
model_1 = LogisticRegression()
model_1.fit(x_train, y_train)
y_pred = model_1.predict(x_test)

# To find F-score
log_acc = model_1.score(x_test, y_test)
log_f1 = f1_score(y_test, y_pred)
```

# Support Vector Machine

```python
model_2 = SVC()
model_2.fit(x_train, y_train)
y_pred = model_2.predict(x_test)

svm_acc = model_2.score(x_test, y_test)
svm_f1 = f1_score(y_test, y_pred)
```

# The Prediction

```python
print(f'Logistic Regression:\nAccuracy: {log_acc}\nF1 Score: {log_f1}\n')
print(f'Support Vector Machine:\nAccuracy: {svm_acc}\nF1 Score: {svm_f1}\n')
```
