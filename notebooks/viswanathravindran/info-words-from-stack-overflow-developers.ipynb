## Gender of the Responders ?

```python
# Change Genders
def genders(row):
    if row['Gender'] == 'Male':
        return "Male"
    elif row['Gender'] == 'Female':
        return "Female"
    else:
        return "Not Available"

survey_results['Gender'] = survey_results.apply(genders, axis = 1).astype(str)
```
