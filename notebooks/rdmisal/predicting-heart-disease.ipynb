But first lets apply the model and check for accuracy.

Splits the data for train and test.

```python
from sklearn.model_selection import train_test_split
train=data.drop('target',axis=1)
test=data['target']
from sklearn.preprocessing  import StandardScaler
scaler=StandardScaler()
scaler.fit(train)
x_train,x_test,y_train,y_test=train_test_split(train,test,test_size=0.33)
```

Apply the naive bayes BernoulliNB classifier and check for the preicted result.

```python
from sklearn.naive_bayes import BernoulliNB
model=BernoulliNB()
model.fit(x_train,y_train)
model.predict(x_test)
```

Now Lets check with accuracy score of the model

```python
from sklearn.metrics import accuracy_score, confusion_matrix
predict=model.predict(x_test)
print("Accuracy score is ", accuracy_score(y_test,predict))
```

Which is really good compared to amount of data.
Lets check with confusion matrix.

```python
confusion_matrix(y_test,predict)
```

Lets check with complete report.

```python
from sklearn.metrics import precision_score, classification_report
classification_report(y_test,predict)
```
