**Changing "Dadra and Nagar Haveli and Daman and Diu" to "DNHDD"**

```python
data["State/UTs"].replace("Dadra and Nagar Haveli and Daman and Diu", "DNHDD", inplace=True)
print(data["State/UTs"].unique())
```

**Hoverable Covid-19 Statwise Stats**

```python

fig = px.bar(data, x='State/UTs', y='Total Cases',
             hover_data=['Active', 'Discharged', "Deaths"], color='State/UTs',
             height=600,
             title="All Statewise Covid-19 Stats for India (Depicts Total Cases per State)")
fig.update_xaxes(tickangle=315)
fig.show()
```

```python
# Calculating Overall Totals:
total_cases = data["Total Cases"].sum()
total_active = data["Active"].sum()
total_discharged = data["Discharged"].sum()
total_deaths = data["Deaths"].sum()
```

```python
# Creating a DataFrame of the computed Totals:
total_dict = {"Total": ['Total Cases', 'Total Active', 'Total Discharged', 'Total Deaths'], 
              "Values": [total_cases, total_active, total_discharged, total_deaths]
             }
total_df = pd.DataFrame(data=total_dict)
total_df
```

```python
fig = px.pie(total_df, names='Total', values='Values', color='Total', title="Covid-19 Totals")
fig.show()
```
