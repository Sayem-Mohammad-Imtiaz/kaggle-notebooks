```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

selected_country_1 = 'Taiwan'
selected_country_2 = 'Taiwan'

df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1 = df1[df1['country']==selected_country_1]
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2 = df2[df2['country']==selected_country_2]
df1.index = pd.to_datetime(df1['date'])
df2.index = pd.to_datetime(df2['date'])
```

# Figure 1
It shows a recent surge in the number of infected people.

```python
fig,ax = plt.subplots(figsize=(10,5))
ax.plot(df1.index,df1['daily_new_cases'],'r',label='Cases')
ax1 = ax.twinx()
ax1.plot(df2.index,df2['daily_vaccinations'],'g',label='Vaccinations')
ax.legend()
ax1.legend(loc='center left')
plt.show()
```

# Figure 2
The more detailed diagram shows imply that something has happened from May 14th to May 17th.

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_new_cases = go.Scatter(x=df1.index,y=df1['daily_new_cases'],yaxis='y1',name='Daily confirmed cases')
daily_vaccinations = go.Scatter(x=df2.index,y=df2['daily_vaccinations'],yaxis='y2',name='Daily vaccinations')

layout_obj = go.Layout(title='Covid vs. Vaccines',xaxis=dict(title ='Date'),yaxis=dict(title='Daily cases'),yaxis2=dict(title='Vaccinations',side='right',overlaying='y'))
fig = go.Figure(data=[daily_new_cases,daily_vaccinations],layout=layout_obj)
iplot(fig)

fig.write_html('outcome.html')
```

# WORLD RANKING
From the table below, Taiwan is ranked 186th in the total number of cases ranking, and the vaccination rate is 180th.
(Regarding the vaccination rate, if it was people_fully_vaccinated_per_hundred, Taiwan became NaN, so I verified by total_vaccinations_per_hundred.）


```python
df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1 = df1.groupby('country').sum()['daily_new_cases'].sort_values(ascending=False)
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2 = df2.groupby('country').max()['total_vaccinations_per_hundred'].sort_values(ascending=False)
print(df1.rank(method='min',ascending=False).to_string())
print(df2.rank(method='min',ascending=False).to_string())
```
