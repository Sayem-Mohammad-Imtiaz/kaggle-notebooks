```python
#Machine Learning
```

```python
#dummie values
ca = pd.get_dummies(df['ca'], drop_first=True)
slope = pd.get_dummies(df['slope'], drop_first = True)
thal = pd.get_dummies(df['thal'], drop_first=True)
```

```python
frames = [df, ca, slope, thal]
df = pd.concat(frames, axis=1)
df.head()
```

```python
df = df.drop('ca', axis=1)
df = df.drop('thal', axis=1)
df = df.drop('slope', axis=1)
```

```python
X = df.drop('target', axis=1)
y = df[['target']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
X_train_ss = ss.fit_transform(X_train)
#X_train_ss
X_test_ss = ss.fit_transform(X_test)
```

```python
lr = LogisticRegression()
lr.fit(X_train_ss, y_train)
score = lr.score(X_train_ss, y_train)
print("Accuracy score for Logestic_Regression",score)
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train_ss, y_train)
score = dtc.score(X_train_ss, y_train)
print("Accuracy score for Decision_Tree_Classifier", score)
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train_ss, y_train)
score = rfc.score(X_train_ss, y_train)
print("Accuracy score for Random_Forest_classifier", score)
```

```python
svc = SVC()
svc.fit(X_train_ss, y_train)
score = svc.score(X_train_ss, y_train)
print("Accuracy score for SVC", score)
```

```python
navie = GaussianNB()
navie.fit(X_train_ss, y_train)
score = navie.score(X_train_ss, y_train)
print("Accuracy score for Navie Bayes", score)
```

```python
knn = KNeighborsClassifier()
knn.fit(X_train_ss, y_train)
score = knn.score(X_train_ss, y_train)
print("Accuracy score for KNN:", score)
```

```python
lr_pred = lr.predict(X_test_ss)
dtc_pred = dtc.predict(X_test_ss)
rfc_pred = rfc.predict(X_test_ss)
svc_pred = svc.predict(X_test_ss)
navie_pred = navie.predict(X_test_ss)
knn_pred = knn.predict(X_test_ss)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
lr_cm = (confusion_matrix(y_test, lr_pred))
dtc_cm =(confusion_matrix(y_test, dtc_pred))
rfc_cm =(confusion_matrix(y_test, rfc_pred))
svc_cm =(confusion_matrix(y_test, svc_pred))
navie_cm =(confusion_matrix(y_test, navie_pred))
knn_cm =(confusion_matrix(y_test, knn_pred))
```

```python
plt.figure(figsize=(24,12))

plt.suptitle("Confusion Matrixes",fontsize=24)
plt.subplots_adjust(wspace = 0.4, hspace= 0.4)

plt.subplot(2,3,1)
plt.title("Logistic Regression Confusion Matrix")
sns.heatmap(lr_cm,annot=True,cmap="Blues",fmt="d",cbar=False, annot_kws={"size": 24})

plt.subplot(2,3,2)
plt.title("K Nearest Neighbors Confusion Matrix")
sns.heatmap(knn_cm,annot=True,cmap="Blues",fmt="d",cbar=False, annot_kws={"size": 24})

plt.subplot(2,3,3)
plt.title("Support Vector Machine Confusion Matrix")
sns.heatmap(svc_cm,annot=True,cmap="Blues",fmt="d",cbar=False, annot_kws={"size": 24})

plt.subplot(2,3,4)
plt.title("Naive Bayes Confusion Matrix")
sns.heatmap(navie_cm,annot=True,cmap="Blues",fmt="d",cbar=False, annot_kws={"size": 24})

plt.subplot(2,3,5)
plt.title("Decision Tree Classifier Confusion Matrix")
sns.heatmap(dtc_cm,annot=True,cmap="Blues",fmt="d",cbar=False, annot_kws={"size": 24})

plt.subplot(2,3,6)
plt.title("Random Forest Confusion Matrix")
sns.heatmap(rfc_cm,annot=True,cmap="Blues",fmt="d",cbar=False, annot_kws={"size": 24})

plt.show()
```
