**3) Which year had the most sales in worldwide**

```python

data_year = data.groupby(by=['Year'])['Global_Sales'].sum()
data_year = data_year.reset_index()
```

```python
plt.figure(figsize=(15, 10))
bar6=sns.barplot(x='Year', y="Global_Sales", data=data_year)
bar6.set_title('Figure 6: Which year had the most sales in worldwide',fontsize=20)
plt.xticks(rotation=90)
```

**From Figure 6, we can see that 2008 had the highest global sales record followed by 2009 and 2007.**

**Kinda made me curious whether humans have starting to move on from gaming since the data for later years have delclined? Obviously, we are still into gaming lol.**

**4) Which genre had the most sales in worldwide**

```python
data_genre = data.groupby(by=['Year','Genre'])['Global_Sales'].sum()
data_genre = data_genre.reset_index()
```

```python
year_sale = data_genre.groupby(by=['Year'])['Global_Sales'].transform(max) == data_genre['Global_Sales']
year_sale_max = data_genre[year_sale].reset_index(drop=True)

genre = year_sale_max['Genre']
```

```python
plt.figure(figsize=(30, 18))
g = sns.barplot(x='Year', y='Global_Sales', data=year_sale_max)
g.set_title('Figure 6.1: Best genre for particular year',fontsize=20)
index = 0
for value in year_sale_max['Global_Sales']:
    g.text(index, value + 1, str(genre[index] + '----' +str(round(value, 2))), color='#000', size=14, rotation= 90, ha="center")
    index += 1

plt.xticks(rotation=90)
plt.show()
```

**As we can see from figure 6.1, action game genre has been on demand since 2001 - 2016.**

**6) SALES-REGION COMPARISON BY GENRE**

```python
data.head(10)
```

```python
# Let comparison be as comp from this point onwards

comp_genre = data[['Genre','NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
comp_heatmap = comp_genre.groupby(by=['Genre']).sum()
```

```python
plt.figure(figsize=(15, 10))
sns.set(font_scale=1)
sns.heatmap(comp_heatmap, annot=True, fmt = '.1f')


plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

**From table above, we can see Action genre has the most sales record in North America, Europe and Other region. However in Japan, role-playing genre has the most sales record than the rest.**

```python
comp_table = comp_heatmap.reset_index()
comp_table = pd.melt(comp_table, id_vars=['Genre'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], var_name='Sale_Area', value_name='Sale_Price')
comp_table.head()

```

```python
plt.figure(figsize=(15, 10))
bar8 = sns.barplot(x='Genre', y='Sale_Price', hue='Sale_Area', data=comp_table)
bar8.set_title('Figure 8: Sales record for each genres in every region comparison',fontsize=20)
```

**Figure 8 shows that North America has the most sales compared to other regions for almost every genres.**

**SALES-REGION COMPARISON BY MANUFACTURER**

```python
comp_manufac = data[['Manufacturer','NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
comp_manufac.head() 
```

```python
comp_manufac = comp_manufac.groupby(by=['Manufacturer']).sum().reset_index()
```

```python
comp_table = pd.melt(comp_manufac, id_vars=['Manufacturer'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], var_name='Sale_Area', value_name='Sale_Price')
comp_table.head()
```

```python
plt.figure(figsize=(30, 15))
bar9 = sns.barplot(x='Manufacturer', y='Sale_Price', hue='Sale_Area', data=comp_table)
bar9.set_title('Figure 9: Sales record for each manufacturers in every region comparison',fontsize=30)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.show()
```

**Figure 9 shows Nintendo has the highest sales record in North America than other regions. However, that does not mean Nintendo is the most on demand. This is because if we take a look closely at Sony Playstation data, we can see for both Europe and Other sales record have higher stats than the other manufacturers. Which is why from the previous studies, we can conclude that Sony Playstation is the next gen console on demand.**
