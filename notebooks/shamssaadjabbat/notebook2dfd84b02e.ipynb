# preprocessed in training file

#   Missing value of trine

```python
remove_null=SimpleImputer(strategy='most_frequent')
x_remove_null=remove_null.fit_transform(data_train)
x_remove_null
```

```python
cls_train=pd.DataFrame(x_remove_null,columns=data_train.columns,index=data_train.index)
cls_train.head()
```

```python
cls_train.isna().sum()
```

```python
def create_dict(column):
  values = column.unique()
  dictionary = {}
  for index, value in enumerate(values):
    dictionary[value] = index

  return dictionary
```

```python
city=create_dict(cls_train.city)
cls_train.city.replace(city,inplace=True)
```

```python
cls_train.city_development_index=cls_train.city_development_index.astype(float)
```

```python
gender=create_dict(cls_train.gender)
cls_train.gender.replace(gender,inplace=True)
```

```python
relevent_experience=create_dict(cls_train.relevent_experience)
cls_train.relevent_experience.replace(relevent_experience,inplace=True)
```

```python
enrolled_university=create_dict(cls_train.enrolled_university)
cls_train.enrolled_university.replace(enrolled_university,inplace=True)
```

```python
major_discipline=create_dict(cls_train.major_discipline)
cls_train.major_discipline.replace(major_discipline,inplace=True)
```

```python
experience=create_dict(cls_train.experience)
cls_train.experience.replace(experience,inplace=True)
```

```python
company_type=create_dict(cls_train.company_type)
cls_train.company_type.replace(company_type,inplace=True)
```

```python
company_size=create_dict(cls_train.company_size)
cls_train.company_size.replace(company_size,inplace=True)
```

```python
last_new_job=create_dict(cls_train.last_new_job)
cls_train.last_new_job.replace(last_new_job,inplace=True)
```

```python
education_level=create_dict(cls_train.education_level)
cls_train.education_level.replace(education_level,inplace=True) 
```

```python
cls_train.training_hours=cls_train.training_hours/336
```

```python
cls_train.training_hours=cls_train.training_hours.astype(float)
```

```python
cls_train.drop(['enrollee_id'], axis=1, inplace=True)
```

```python
cls_train.head()
```

```python
cls_train['target']=cls_train['target'].astype(int)
```

```python
cls_train.info()
```

```python
y=cls_train['target']
x=cls_train.drop(['target'], axis = 1)
```

```python
print('x:',x.shape)
print('y:',y.shape)
```

# Split dataset

```python
from sklearn import datasets, model_selection, svm, metrics
```

```python
x_train, x_test, y_train, y_test=model_selection.train_test_split(x,y,train_size=0.1, random_state=42, stratify=y)
```

```python
x_train.shape, y_train.shape, y_test.shape,x_test.shape
```

# Models

# Logistic Regression

```python
Logistic_Regression = linear_model.LogisticRegression(random_state=42)
Logistic_Regression.fit(x_train, y_train)
y_pred = Logistic_Regression.predict(x_test)
metrics.accuracy_score(y_test, y_pred)
```

```python
svc_model= svm.SVC(kernel='linear')
svc_model.fit(x_train, y_train)
y_pred = svc_model.predict(x_test)
print("Accuracy Score:", metrics.accuracy_score(y_test, y_pred))
```

# RandomForest

```python
random_forest = ensemble.RandomForestClassifier()
random_forest.fit(x_train, y_train)
y_pred_forest = random_forest.predict(x_test)
metrics.accuracy_score(y_test, y_pred_forest)
```

# DecisionTree

```python
tree = tree.DecisionTreeClassifier()
tree.fit(x_train, y_train)
y_pred_tree = tree.predict(x_test)
metrics.accuracy_score(y_test, y_pred_tree)
```

# Neural Networks in TensorFlow

```python
import tensorflow as tf
from tensorflow import keras
import IPython
import kerastuner as kt
```

```python
y_tf=cls_train['target']
x_tf=cls_train.drop(['target'], axis = 1)
x_train_tf, x_test_tf, y_train_tf, y_test_tf=model_selection.train_test_split(x_tf,y_tf,train_size=0.1, random_state=42, stratify=y_tf)
x_train_tf=x_train_tf.to_numpy()
x_test_tf=x_test_tf.to_numpy()
y_train_tf=y_train_tf.to_numpy()
y_test_tf=y_test_tf.to_numpy()
print('x_test',x_test_tf.shape)
print('y_test',y_test_tf.shape)
print('x_train',x_train_tf.shape)
print('y_train',y_train_tf.shape)
print('type x_test',type(x_test_tf))
print('type y_test',type(y_test_tf))
print('type x_train',type(x_train_tf))
print('type y_train',type(y_train_tf))
```

```python
model_1 = tf.keras.Sequential([
                                tf.keras.layers.Input(12),
                                tf.keras.layers.Dense(16,activation=tf.nn.leaky_relu),
                                tf.keras.layers.Dropout(0.2),
                                tf.keras.layers.Dense(16,activation=tf.nn.leaky_relu),
                                tf.keras.layers.Dropout(0.2),
                                tf.keras.layers.Dense(8,activation=tf.nn.leaky_relu),
                                tf.keras.layers.Dense(1, name='output_layer',activation=tf.nn.sigmoid)
])
model_1.summary()
```

```python
model_1.compile(optimizer=tf.optimizers.Adam(), loss=tf.losses.BinaryCrossentropy(),metrics=['acc'])
```

```python
history_1=model_1.fit(x=x_train_tf,y=y_train_tf, epochs=25, validation_data=(x_test_tf, y_test_tf))
```

```python
predict=model_1.predict(x_test_tf)
tf.reduce_mean(tf.keras.metrics.binary_accuracy(y_test_tf,predict))
```

```python
df_1 = pd.DataFrame(history_1.history)
df_1
```
