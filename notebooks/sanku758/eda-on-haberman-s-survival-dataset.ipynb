### Observations:
These histograms shows that PDF of both Age and Num of nodes collumn are overlapping, lets try cummulative density functions then

# CDF

```python
label = ["pdf of Survival st = 1", "cdf of Survival st = 1", "pdf of Survival st = 2", "cdf of Survival st = 2"]
counts, bin_edges = np.histogram(df[df.Survival_status == 1]['Age'], bins=10, density=True)
pdf = counts / sum(counts)
#print(pdf)
#print(bin_edges)
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf);

counts, bin_edges = np.histogram(df[df.Survival_status == 2]['Age'], bins=10, density=True)
pdf = counts / sum(counts)
#print(pdf)
#print(bin_edges)
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf);
plt.legend(label);
```

```python
label = ["pdf of Survival st = 1", "cdf of Survival st = 1", "pdf of Survival st = 2", "cdf of Survival st = 2"]
counts, bin_edges = np.histogram(df[df.Survival_status == 1]['Num_of_pos_axillary_nodes'], bins=10, density=True)
pdf = counts / sum(counts)
#print(pdf)
#print(bin_edges)
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf);

counts, bin_edges = np.histogram(df[df.Survival_status == 2]['Num_of_pos_axillary_nodes'], bins=10, density=True)
pdf = counts / sum(counts)
#print(pdf)
#print(bin_edges)
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf);
plt.legend(label);
```

```python
label = ["pdf of Survival st = 1", "cdf of Survival st = 1", "pdf of Survival st = 2", "cdf of Survival st = 2"]
counts, bin_edges = np.histogram(df[df.Survival_status == 1]['Year_of_treatment'], bins=10, density=True)
pdf = counts / sum(counts)
#print(pdf)
#print(bin_edges)
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf);

counts, bin_edges = np.histogram(df[df.Survival_status == 2]['Year_of_treatment'], bins=10, density=True)
pdf = counts / sum(counts)
#print(pdf)
#print(bin_edges)
cdf = np.cumsum(pdf)

plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf);
plt.legend(label);
```

As shown in the plots, both PDF and CDF tends to overlap.
