```python
df = pd.read_csv("/kaggle/input/suicide-rates-overview-1985-to-2016/master.csv")
df
```

```python
df.info()
```

```python
df['age'].replace('5-14 years','05-14 years',inplace=True)
df
```

```python
Age_range = df.groupby('age').agg({'suicides_no':np.sum})
Age_range
```

```python
import matplotlib.pyplot as plt
plt.plot(Age_range,marker='o')
plt.ylim([10000,2500000])
plt.xticks(rotation=45)
plt.title('We can clearly see that no. of suicide cases are maximum in age range 35-54 years\n and they are 2452141',pad=50,fontsize=25)
```

```python
plt.pie(Age_range['suicides_no'],labels=Age_range.index)
```

```python
year_wise_df = df.groupby(['year','age']).agg({'suicides_no':np.sum})
year_wise_df.head(10)
```

```python
yr = pd.DataFrame(df['year'].value_counts()).sort_index()
yr.index
```

```python
yr.plot(kind ='bar')
plt.title('Suicides recorded per year',pad=25,fontsize=25)
```

```python
age_range_05to14 = year_wise_df.loc[[(2000,'05-14 years'), (2001,'05-14 years'), (2002,'05-14 years'), (2003,'05-14 years'),
                                    (2004,'05-14 years'), (2005,'05-14 years'), (2006,'05-14 years'), (2007,'05-14 years'), 
                                    (2008,'05-14 years'), (2009,'05-14 years'), (2010,'05-14 years'), (2011,'05-14 years'), 
                                    (2012,'05-14 years'), (2013,'05-14 years'), (2014,'05-14 years'), (2015,'05-14 years')]].reset_index()
```

```python
age_range_05to14[age_range_05to14['suicides_no']==min(age_range_05to14['suicides_no'])]
```

```python
plt.plot(age_range_05to14['year'],age_range_05to14['suicides_no'],marker='o')
ymax = 2095
xmax = 2001
plt.title('Number of Suicides commited by age_range: 5 to 14 years',pad=50,fontsize=25)
plt.gca().annotate('2095 is maximum no. of suicides in 2001 \n', xy=(xmax, ymax), xytext=(xmax, ymax+5),
            arrowprops=dict(facecolor='red', shrink=0.05) )
plt.ylim(0,2500)
ymin = 1640
xmin = 2008
plt.gca().annotate('  1640 is minimum no. of suicides in 2008', xy=(xmin, ymin), xytext=(xmin, ymin+30),
            arrowprops=dict(facecolor='red', shrink=0.05) )
plt.xlabel('Year')
plt.ylabel('Number of Suicides')
```
