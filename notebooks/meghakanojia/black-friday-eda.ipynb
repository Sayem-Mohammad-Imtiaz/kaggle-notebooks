```python
X=compare_df[['Product_Category_1','Product_Category_2','Product_Category_3']]
y=compare_df['Purchase']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y)
linear_model=LinearRegression().fit(X_train,y_train)
```

```python
linear_model.score(X_train,y_train)
```

```python
linear_model.score(X_test,y_test)
```

```python
# Polynomial Features alongwith Ridge Regression
from sklearn.linear_model import Ridge
X_poly=PolynomialFeatures(degree=2).fit_transform(X)
X_poly_train,X_poly_test,y_train,y_test=train_test_split(X_poly,y)
Poly_Ridge_model=Ridge(alpha=1.0).fit(X_poly_train,y_train)
```

```python
Poly_Ridge_model.score(X_poly_test,y_test)
```

```python
def Poly(degree):
    X_poly=PolynomialFeatures(degree=degree).fit_transform(X)
    X_poly_train,X_poly_test,y_train,y_test=train_test_split(X_poly,y)
    Poly_Ridge_model=Ridge(alpha=5.0).fit(X_poly_train,y_train)
    print(Poly_Ridge_model.score(X_poly_train,y_train),Poly_Ridge_model.score(X_poly_test,y_test))
```

```python
from sklearn.linear_model import Ridge
for degree in [2,3,4,5,6,8,10,11,12]:
    Poly(degree)
```

```python
def Polyy(degree):
    X_poly=PolynomialFeatures(degree=degree).fit_transform(X)
    X_poly_train,X_poly_test,y_train,y_test=train_test_split(X_poly,y)
    Poly_Ridge_model=Ridge(alpha=8.0).fit(X_poly_train,y_train)
    print(Poly_Ridge_model.score(X_poly_train,y_train),Poly_Ridge_model.score(X_poly_test,y_test))
```

```python
for degree in [2,3,4,5,6,8,10,11,12]:
    Polyy(degree)
```

```python
## Best fit: Polynomial+Ridge --> degree=10 , alpha=8.0
```

```python
def knn(neighbors):
    from sklearn.neighbors import KNeighborsRegressor
    X_train,X_test,y_train,y_test=train_test_split(X,y)
    knn_model=KNeighborsRegressor(n_neighbors=neighbors).fit(X_train,y_train)
    print(knn_model.score(X_train,y_train),knn_model.score(X_test,y_test))
```

```python
for nn in [10,20,25,30]:
    knn(nn)
```

```python
for nn in [35,40,60]:
    knn(nn)
```

```python
for nn in [90]:
    knn(nn)
```
