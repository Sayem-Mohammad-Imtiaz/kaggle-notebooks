Hello everyone!This is my  first notebook.It mainly focuses on data visualisation using python.This dataset mainly focuses on how students get placed for their jobs and whether their salary is dependent on multiple factors like ssc percentage ,hsc percentage etc.Any suggestrions is welcomed here.

```python
#importing lib
import numpy as nm
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
#loading the file
df=pd.read_csv('../input/factors-affecting-campus-placement/Placement_Data_Full_Class.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
#checking for missing values
df.isnull().sum()
#salary has many null values because the students that are not placed cannot get salary.
```

```python
#salary is the dependent variable which is dependent on other variables like ssc percentage,mba percentage etc
#correlation heatmap
#correlation is the degree of relation between  an independent variable and a dependent variable.
plt.subplots(figsize=(15,10))
sns.heatmap(df.corr(),annot=True,linewidths=.5,linecolor='black',fmt='1.1f')
plt.title('correlation heatmap')
plt.show()
```

there  is a strong correlation between degree_p and ssc_p,hsc_p and ssc_p etc

# data visualisation

```python
#relation between gender and salary using  countplot
sns.countplot(data=df,x='gender')
```

```python
#there are more male candidates in this dataset.
```

```python
# relation betwwen gender and salary using boxplot
sns.boxplot(x='gender',y='salary',data=df)
plt.show()
```

```python
#there are a lot of outliers in the salaries of males. the middle line is the median salary and the lower half and the upper half represents the 25th quartile and 75 th quartile respectively
```

# relation between ssc_p and salary

```python
plt.hist(df['ssc_p'],color='salmon')
plt.xlabel('ssc percentage')
plt.ylabel('salary')
plt.show()
```

# relationship between ssc_b and salary

```python
#ssc_b is a categorical variable  
#using barchart
sns.countplot(x='ssc_b',data=df)

plt.show()
```

```python
#relation between gender and status
sns.countplot(data=df,x='status',hue=df['gender'])
#females are less compared to males but mostly they are getting placed.
```

# relationship between hsc_p and salary


```python
#using boxplot
plt.hist(df['hsc_p'],color='salmon')
plt.xlabel('hsc_p')
plt.ylabel('salary')
plt.show()
```

```python
#there is no such relation between hsc_p and salary because the students with mediocre percentage got the highest salaries.
```

# relationship between hsc_b and salary


```python
sns.boxplot(x='hsc_b',y='salary',data=df)
plt.show()
```

```python
#a student studying in central board got the highest salary.But there is no such correlation between boards and salaries because the median salary is mor or less the same.
```

```python
sns.countplot(data=df,x='status',hue=df['hsc_s'])
```

```python
#more students are of commerce field.there is 50% of chance that students of arts field get placed or donot get placed.
```

# mba_p and salary

```python
plt.hist(x=df['mba_p'],color='pink')
plt.xlabel('mba_p')
plt.ylabel('salary')
plt.show()
```

```python
#students with mediocre percentage got good packages(salaries)
```

# mba_p and status

```python
sns.catplot(x='status',y='ssc_p',data=df,kind='swarm',hue='gender')
sns.catplot(x='status',y='hsc_p',data=df,kind='swarm',hue='gender')
sns.catplot(x='status',y='degree_p',data=df,kind='swarm',hue='gender')
```
