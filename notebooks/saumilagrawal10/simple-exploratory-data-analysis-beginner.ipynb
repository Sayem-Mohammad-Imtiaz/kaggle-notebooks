#### There are outliers in case of total_vaccinations_per_hundred. However there is another finding where median is towards left which says the data is right-skewed.

#### Data is Positively Skewed . Mean > Median (Right Side Skewed)

```python
print("Mean total_vaccinations_per_hundred:")
print(df["total_vaccinations_per_hundred"].mean())
print("Median total_vaccinations_per_hundred:")
print(df["total_vaccinations_per_hundred"].median())
```
