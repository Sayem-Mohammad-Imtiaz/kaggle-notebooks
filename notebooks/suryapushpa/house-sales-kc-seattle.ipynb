# Simple Linear Regression

### Importing Libraries

```python
import pandas as pd 
pd.set_option('display.max_rows',None)

import numpy as np

import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

### Collecting Data

```python
data = pd.read_csv("../input/housesalesprediction/kc_house_data.csv")
data.head()
```

### Exploring Data

```python
data.info()
```

```python
data.isnull().sum()
```

### Check for any correlations between variables

```python
corr = data.corr()
plt.figure(figsize=(10,10))
sns.heatmap(corr, linewidths=0.1, cmap='YlGnBu')
plt.show()
```

# Building a Simple Linear Regression Model

### Preparing Data and training set

```python
# Create x and y

featured_col = 'sqft_living'
x = data[featured_col]
y = data['price']

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)

x_train = x_train[:,np.newaxis]
y_train = y_train[:,np.newaxis]
x_test = x_test[:,np.newaxis]
y_test = y_test[:,np.newaxis]
```

### Fit the model, Get the coefficients

```python
reg = LinearRegression()
reg.fit(x_train, y_train)

reg.intercept_, reg.coef_
```

### Predict the values and calculate error

```python
yi_val = reg.coef_ * 1000 + reg.intercept_
print(yi_val)
print(reg.predict([[1000]]))

mse = mean_squared_error(y_test, reg.predict(x_test))
print(np.sqrt(mse))
# Not great

print(reg.score(x_test, y_test))
# Our model is able to predict only 50% of the variability in house prices.
```
