```python
from sklearn.ensemble import RandomForestRegressor
re=RandomForestRegressor()
re.fit(x_train,y_train)
acc=re.score(x_train,y_train)
print(acc)
```

```python
from sklearn.linear_model import LinearRegression
le=LinearRegression()
le.fit(x_train,y_train)
acc=le.score(x_train,y_train)
print(acc)
```

```python
from sklearn.tree import DecisionTreeRegressor
de=DecisionTreeRegressor()
de.fit(x_train,y_train)
acc=de.score(x_train,y_train)
print(acc)
```

```python
re_predict=re.predict(x_train)
de_predict=de.predict(x_train)
le_predict=le.predict(x_train)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```
