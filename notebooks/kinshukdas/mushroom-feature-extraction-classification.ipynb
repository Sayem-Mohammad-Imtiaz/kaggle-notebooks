Here the Mushrooom Classification dataset has ben analyzed, basic preprocessing, feature importance, feature extraction and a classifier has been build upon the data set.

```python
# Importing the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('../input/mushrooms.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# Encoding categoricakl variables

encoder = LabelEncoder()
data = pd.DataFrame()

for col in df.columns:
    data[col] = encoder.fit_transform(df[col])

data.head()
```

```python
sns.countplot(x='class', data=data)
plt.xlabel('Class')
plt.ylabel('Count')
```

Here the feature importance has been calculated using Random Forest Classifier.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
X = data.drop(['class'], axis=1)
y = data['class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = RandomForestClassifier(n_estimators=500)
model.fit(X_train_scaled, y_train)
```

```python
importance = model.feature_importances_
indices = np.argsort(importance)[::-1]
```

```python
feature_labels = list(X_train.columns)
```

```python
import plotly.offline as pyo
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode

init_notebook_mode(connected=True)
```

```python
# Plotting the feature importance

trace = go.Bar(
    x = [feature_labels[i] for i in range(X_train_scaled.shape[1])],
    y = importance[indices]
)

layout = go.Layout(
    {'title': 'Random Forest Feature Importance'},
    xaxis = {'title': 'features'},
    yaxis = {'title': 'feature importance'}
)

fig = go.Figure(data=[trace], layout=layout)
pyo.iplot(fig)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=5)

X_train_reduced = pca.fit_transform(X_train)
X_test_reduced = pca.transform(X_test)
```

```python
pca.explained_variance_ratio_
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train_reduced, y_train)
```

```python
y_pred = lr_model.predict(X_test_reduced)
```

```python
print('Logistic Regression Accuracy: {}'.format(metrics.accuracy_score(y_test, y_pred)))
```

```python
print(metrics.classification_report(y_test, y_pred))
```
