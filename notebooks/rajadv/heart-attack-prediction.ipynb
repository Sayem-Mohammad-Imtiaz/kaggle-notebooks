```python
x = df.drop(['output'], axis=1)
y = df['output']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit_transform(x_train)
sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state = 10)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```

```python
from sklearn.svm import SVC
cl = SVC(kernel='rbf')
cl.fit(x_train,y_train)
```

```python
y_pred2 = cl.predict(x_test)
```

```python
accuracy_score(y_pred2,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
cl2 = RandomForestClassifier(n_estimators=300,criterion="gini",random_state=5,max_depth=100)
cl2.fit(x_train,y_train)
```

```python
y_pred3 = cl2.predict(x_test)
```

```python
accuracy_score(y_pred3,y_test)
```

 **So we got the best accuracy of 0.8524590163934426 with Logistic Regression**
