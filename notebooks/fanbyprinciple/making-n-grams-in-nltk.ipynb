```python
import collections
from nltk import ngrams
```

```python
import pandas as pd

data = pd.read_csv("../input/first-gop-debate-twitter-sentiment/Sentiment.csv")

all_tweets = []
for i in data['text']:
    all_tweets.append(str(i))
```

```python
tweets = "".join(all_tweets)
```

```python
len(tweets)
```

```python
Ngrams = ngrams(tweets, 4)
```

```python
extracted_ngrams = collections.Counter(list(Ngrams))
```

```python
print(extracted_ngrams.most_common(10))
```

Lets put K1 = 1000 most frequent N-grams into a list

```python
K1 = 1000

K1_most_frequent_Ngrams = extracted_ngrams.most_common(K1)
K1_most_frequent_Ngrams_list = [x[0] for x in K1_most_frequent_Ngrams]
```
