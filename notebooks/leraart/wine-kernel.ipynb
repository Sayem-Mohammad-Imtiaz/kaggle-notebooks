**А еще было бы неплохо попробовать случайные леса**

```python
#1 - плохое
#2 - нормальное
#3 - превосходное
#1,2,3 --> плохое
#4,5,6,7 --> нормальное
#8,9,10 --> превосходное
reviews = []
for i in df['quality']:
    if i >= 1 and i <= 3:
        reviews.append('1')
    elif i >= 4 and i <= 7:
        reviews.append('2')
    elif i >= 8 and i <= 10:
        reviews.append('3')
df['Reviews'] = reviews
```

```python
x = df.iloc[:,:11]
y = df['Reviews']
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
```

```python
pca_new = PCA(n_components=8)
x_new = pca_new.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_new, y, test_size = 0.4)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
rf.fit(x_train, y_train)
```

```python
rf_predict=rf.predict(x_test)
```

```python
print(rf.feature_importances_)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
acc_score = accuracy_score(y_test, rf_predict)
print(acc_score)
```

**Вот это да! 98%\ Вывод: случайные леса - лучший вариант!**
