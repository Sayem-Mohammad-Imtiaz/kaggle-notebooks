```python
data = pd.read_csv('/kaggle/input/mushroom-classification/mushrooms.csv')
```

```python
pd.set_option("display.max_columns",25)
```

```python
data.head()
```

```python
target = data['class'].copy()
```

```python
data.drop('class', axis=1, inplace=True)
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split

train_x, test_x, train_y, test_y = train_test_split(data, target, test_size=0.2, random_state=42)
```

```python
train_x.info()
```

```python
train_x.nunique()
```

```python
test_x.nunique()
```

I will use One Hot Encoder for Encoding the data.

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
train_x_enc = ohe.fit_transform(train_x)
```

```python
train_x_enc[0,:].todense()
```

```python
test_x_enc = ohe.transform(test_x)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

rfc = RandomForestClassifier()
dtc = DecisionTreeClassifier(random_state=0)
```

```python
dtc.fit(train_x_enc, train_y)
```

```python
predict = dtc.predict(test_x_enc)
predict
```

```python
from sklearn.metrics import confusion_matrix, plot_confusion_matrix, accuracy_score, f1_score, classification_report

print(confusion_matrix(test_y, predict))
```

```python
accuracy_score(test_y, predict)
```

```python
f1_score(test_y, predict, average='weighted')
```

```python
print(classification_report(test_y, predict))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(test_y, predict)
plt.figure(figsize = (5,5))
sns.heatmap(cm, annot=True, fmt='d')  # d for decimal
plt.show()
```
