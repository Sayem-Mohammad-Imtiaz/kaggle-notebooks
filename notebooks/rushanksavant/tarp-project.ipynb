```python
import pandas as pd 
```

```python
data= pd.read_csv("../input/prediction-of-asteroid-diameter/Asteroid_Updated.csv")
data.info
```

```python
data
```

```python
data.isnull().sum()
```

```python
df= data[['a', 'e', 'i', 'om', 'w', 'q', 'ad', 
          'per_y', 'data_arc', 'n_obs_used','neo',
       'diameter', 'moid', 'class', 'n', 'per', 'ma']]
```

```python
df = df[df['diameter'].notna()]
df= df.reset_index(drop= True)
```

```python
df.isnull().sum()
```

```python
df = df[df['data_arc'].notna()]
df= df.reset_index(drop= True)
```

```python
df.isnull().sum()
```

```python
# Converting diameter data from str to float
df['diameter']= df['diameter'].astype(float)
```

```python
# Categorical columns
for i in df.columns:
    if type(df[i][5]) is str:
        print(i)
```

```python
#Printing the unique values(categories) of each categorical feature
for column in ['neo', 'class']:
    print(column, df[column].unique())
```

```python
# Applying label encoder on whole dataset
# Import label encoder
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
df['neo']= label_encoder.fit_transform(df['neo'])
df['class']= label_encoder.fit_transform(df['class'])
```

```python
df
```

```python
# Selecting target and independent features
X = df[['a', 'e', 'i', 'om', 'w', 'q', 'ad', 'per_y', 'data_arc', 
        'n_obs_used','neo', 'moid', 'class', 'n', 'per', 
        'ma']]  #independent columns
y = df['diameter']    #target column i.e price range
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.25)
```

```python
from sklearn.metrics import r2_score
```

### Linear Regression

```python
# Fitting Linear Regression to the Dataset
from sklearn.linear_model import LinearRegression
lin_reg= LinearRegression()
lin_reg.fit(X_train, Y_train)

Y_pred_lr= lin_reg.predict(X_test)
print(Y_pred_lr)
```

```python
coefficient_of_dermination_lr = r2_score(Y_test, Y_pred_lr)
print(coefficient_of_dermination_lr)

accu= lin_reg.score(X_train, Y_train)
print('Accuracy for the trained model is:' + str(accu*100) + '%')
```

### k-Nearest Neighbours 

```python
# Fitting KNN Regression to the Dataset
from sklearn.neighbors import KNeighborsRegressor

neigh = KNeighborsRegressor(n_neighbors=3)
neigh.fit(X_train,Y_train)

Y_pred_neigh = neigh.predict(X_test)
```

```python
coefficient_of_dermination_knn = r2_score(Y_test, Y_pred_neigh)
print(coefficient_of_dermination_knn)

accu= neigh.score(X_train, Y_train)
print('Accuracy for the trained model is:' + str(accu*100) + '%')
```

### Random Forest  

```python
# Fitting Random Forest Regressor Regression to the Dataset
from sklearn.ensemble import RandomForestRegressor
regressor_rr= RandomForestRegressor(n_estimators=200, random_state= 0) #random estiators= Number of tress
regressor_rr.fit(X_train, Y_train)

Y_pred_rr = regressor_rr.predict(X_test)
print(Y_pred_rr)
```

```python
coefficient_of_dermination_rr = r2_score(Y_test, Y_pred_rr)
print(coefficient_of_dermination_rr)

accu= regressor_rr.score(X_train, Y_train)
print('Accuracy for the trained model is:' + str(accu*100) + '%')
```
