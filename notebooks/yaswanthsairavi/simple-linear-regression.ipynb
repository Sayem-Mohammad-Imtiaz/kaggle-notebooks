```python
#importing all the required libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
# Load the salary dataset
salarydata = pd.read_csv("../input/salary-data-simple-linear-regression/Salary_Data.csv")
# assigning colums to X and y
X, y = salarydata.iloc[:, :-1], salarydata.iloc[:, [-1]]

# Split the data into training/testing sets using train_test_split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.70)


# Create linear regression object
regr = linear_model.LinearRegression()

# Training the model using the training sets
regr.fit(X_train,y_train)

# Making predictions using the testing set
y_pred = regr.predict(X_test)

# this prints the cofficients for the regression equation
print('Coefficients: \n', regr.coef_)
# The returns the mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test,y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y_test, y_pred))
#returns the rsme value 
rmse_val = np.sqrt(mean_squared_error(y_test,y_pred))
print(rmse_val)

# Plot outputs
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test,y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())
plt.show()
```

```python

# Load the salary dataset
salarydata = pd.read_csv("../input/salary-data-simple-linear-regression/Salary_Data.csv")
#here I am doing an extra operation i.e removing the duplicate rows , It helps to reduce the mean square error and improve the performance of algorithm
salarydata = salarydata.drop_duplicates(subset=['YearsExperience'],keep="last")

# assigning colums to X and y
X, y = salarydata.iloc[:, :-1], salarydata.iloc[:, [-1]]

# Split the data into training/testing sets using train_test_split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.70)


# Create linear regression object
regr = linear_model.LinearRegression()

# Training the model using the training sets
regr.fit(X_train,y_train)

# Making predictions using the testing set
y_pred = regr.predict(X_test)

# this prints the cofficients for the regression equation
print('Coefficients: \n', regr.coef_)
# The returns the mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test,y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y_test, y_pred))
#returns the rsme value 
rmse_val = np.sqrt(mean_squared_error(y_test,y_pred))
print(rmse_val)

# Plot outputs
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test,y_pred, color='blue', linewidth=3)

plt.xticks(())
plt.yticks(())

plt.show()
```
