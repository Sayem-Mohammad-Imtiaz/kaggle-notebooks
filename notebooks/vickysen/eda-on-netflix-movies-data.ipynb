## IMDb Score Analysis

```python
df['IMDB Score'].head(3)
```

```python
df['IMDB Score'].describe().T
```

#### Here we can see the maximum given rating is 9.0 and minimum rating is 2.5, and average rating is 6.27 and median of all the rating is 6.35, since median is bigger than mean score so our data may have outlier. in order to visualize that we will plot a histogram for that

```python
# plottting a histogram with IMDB scre
sns.histplot(x=df['IMDB Score'],color='blue',data=df)
plt.show()
```

#### we can see that our data is left skewed, negatively distributed toward left

```python
# drawing a box plot to visualize the outliers
import warnings
warnings.filterwarnings('ignore')
sns.boxplot(df['IMDB Score'])
```

```python
df.groupby("Genre")["IMDB Score"].nlargest(20)
```

```python
top_10_ratings_by_genre=df.groupby("Genre")["IMDB Score"].mean().nlargest(10)
top_10_ratings_by_genre
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=top_10_ratings_by_genre.index,y=top_10_ratings_by_genre.values,color='blue')
plt.xlabel('Genre',fontdict = { 'color':  'darkblue', 'weight': 'normal', 'size': 12})
plt.ylabel('Rating',fontdict = {'family': 'times new roman','size':12})
plt.xticks(rotation=90)
plt.show()
```

```python
lowest_10_ratings_by_genre = df.groupby("Genre")["IMDB Score"].mean().nsmallest(10)
lowest_10_ratings_by_genre
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=lowest_10_ratings_by_genre.index,y=lowest_10_ratings_by_genre.values,color='blue')
plt.xlabel('Genre',fontdict = { 'color':  'darkblue', 'weight': 'normal', 'size': 12})
plt.ylabel('Rating',fontdict = {'family': 'times new roman','size':12})
plt.xticks(rotation=90)
plt.show()
```

#### heist thriller film has the lowest rating
