```python
# Extracting dependent and independent varirable for training
X=diabetes_df.drop(['Outcome'],axis=1)
y=diabetes_df.Outcome
```

```python
# Now before feeding our data to training algorithm we need to do some preprocessing
# So, here we do Standard scaling first

from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
```

```python
# scaling input data using StandardScaler & then converting it into a dataframe
X =  pd.DataFrame(ss.fit_transform(diabetes_df.drop(["Outcome"],axis = 1)),
        columns=['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age'])
```

```python
# prining top 5 rows of scaled input data
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
```

```python
# importing KNeighborsClassifier for model training 
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
```

```python
# fitting the model
knn.fit(X,y)
```

```python
# getting model traning score
knn.score(X_train,y_train)
```

```python
# getting model test score
knn.score(X_test,y_test)
```

```python
# definind prediction vector
y_pred=knn.predict(X_test)
```
