```python
import pandas as pd
import numpy as np
import io
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.metrics import accuracy_score

# stroke_df = pd.read_csv(io.BytesIO(myfile['healthcare-dataset-stroke-data (1).csv']))
# stroke = pd.read_csv(io.BytesIO(myfile['healthcare-dataset-stroke-data (1).csv']))

stroke_df = pd.read_csv('/kaggle/input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')
stroke = pd.read_csv('/kaggle/input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')

stroke_df = pd.get_dummies(stroke_df, columns=['gender','ever_married','work_type','Residence_type','smoking_status']) #one hot encoding을 통해 object type column -> int type 
stroke_df.drop('id', axis=1, inplace=True) #prediction에 전혀 영향이 없는, 오히려 성능을 저하시키는 column drop
df_bmi = stroke_df[stroke_df['bmi'].isnull()] #bmi가 NaN인 행만 추출
df_bmi = df_bmi.reset_index(drop=True)
df_bmi.drop('bmi', axis=1, inplace=True)
stroke_df = stroke_df.dropna(axis=0)

y_target = stroke_df['bmi']
x_data = stroke_df.drop(['bmi'], axis=1, inplace=False)
x_train, x_test, y_train, y_test = train_test_split(x_data, y_target, test_size=0.2, shuffle=True, random_state=11)
linear = LinearRegression()
linear.fit(x_train,y_train)
pred = linear.predict(df_bmi)
pred_list = pred.tolist()

bmi = stroke['bmi']
bmi_list = bmi.tolist()

for i in range(0, len(bmi_list)):
  if np.isnan(bmi_list[i]):
    bmi_list[i] = pred_list.pop(0)
  else:
    next

stroke['bmi_new'] = bmi_list
stroke.drop('bmi', axis=1, inplace=True)
stroke = pd.get_dummies(stroke, columns=['gender','ever_married','work_type','Residence_type','smoking_status'])

target = stroke['stroke']
feature = stroke.drop('stroke', axis=1, inplace=False)
f_train, f_test, t_train, t_test = train_test_split(feature, target, test_size=0.2, shuffle=True)
logistic = LogisticRegression()
logistic.fit(f_train, t_train)
prediction = logistic.predict(f_test)
print("accuracy:{}".format(accuracy_score(prediction, t_test)))
```
