We are using Label Encoder to assign values to unique categories in 'Stay' column


```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df_train['Stay']=le.fit_transform(df_train['Stay'])
```

Number of unique categories for each column

```python
cols=df_train.columns
for col in cols:
    print('Number of unique categories for ',col,'are:',df_train[col].nunique())
```

Dropping 'case_id' and 'patientid' columns as they have more than one unique categories

```python
df_train.drop(['case_id','patientid'],axis=1,inplace=True)
```

```python
df_train.isna().sum()
```

'Bed Grade' column and 'City_Code_Patient' column have 113 and 4532 null values. Replace 'Bed Grade' null values by '1' and 'City_Code_Patient' null values by '0'.  

```python
df_train['Bed Grade']=np.where(df_train['Bed Grade'].isna(),1.0,df_train['Bed Grade'])
```

```python
df_train['City_Code_Patient']=np.where(df_train['City_Code_Patient'].isna(),0.0,df_train['City_Code_Patient'])
```
