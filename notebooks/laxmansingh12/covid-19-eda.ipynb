## Checking the null values

```python
nan=data.isnull().sum().to_frame()
sns.heatmap(nan,annot=True,fmt='d',cmap="BuPu")
```

## Separating the Continous and Categorical Variable

```python
data_cont=data.drop("State/UTs",axis=1)
data_cat=data["State/UTs"]
```

## Stastical Analysis

```python
data.describe().transpose()
```

##  Graphical Analysis Of Dataset

```python
print("Histogram Plots of Continous Variable")
for i in data_cont.columns:
    plt.title(i)
    plt.hist(data_cont[i])
    plt.show()
```

## Total Corona Cases of all States/UTs

```python
Cases=data[["Total Cases",'State/UTs']]
cases_sort=Cases.sort_values(by="Total Cases",ascending=False)
px.histogram(cases_sort,x='State/UTs',y='Total Cases',color='State/UTs')
```

## Active Corona Cases of all States/UTs

```python
act_Cases=data[["Active",'State/UTs']]
cases_act=act_Cases.sort_values(by="Active",ascending=False)
px.histogram(cases_act,x='State/UTs',y='Active',color='State/UTs')
```

## Discharged Patients in all States/UTs

```python
discharge_Cases=data[["Discharged",'State/UTs']]
cases_dis=discharge_Cases.sort_values(by="Discharged",ascending=False)
px.histogram(cases_dis,x='State/UTs',y='Discharged',color='State/UTs')
```

## No. of Death in States/Uts

```python
death=data[["Deaths",'State/UTs']]
death_sort=death.sort_values(by="Deaths",ascending=False)
px.histogram(death_sort,x='State/UTs',y='Deaths',color='State/UTs')
```

## Active Ratio of all States/UTs

```python
ratio_act=data[["Active Ratio (%)",'State/UTs']]
ratio=ratio_act.sort_values(by="Active Ratio (%)",ascending=False)
px.histogram(ratio,x='State/UTs',y='Active Ratio (%)',color='State/UTs')
```

## Discharge Ratio of all States/UTs

```python
ratio_dis=data[["Discharge Ratio (%)",'State/UTs']]
ratio_sort=ratio_dis.sort_values(by="Discharge Ratio (%)",ascending=False)
px.histogram(ratio_sort,x='State/UTs',y='Discharge Ratio (%)',color='State/UTs')
```

## Death Ratio Of all States/UTs

```python
ratio_death=data[["Death Ratio (%)",'State/UTs']]
ratio_sort_death=ratio_death.sort_values(by="Death Ratio (%)",ascending=False)
px.histogram(ratio_sort_death,x='State/UTs',y='Death Ratio (%)',color='State/UTs')
```

## Active And Discharge Ratio In Individual States/UTs

```python
px.scatter_3d(data,x="State/UTs",y="Active Ratio (%)",z="Discharge Ratio (%)",color='State/UTs')
```

## Correlation between the Variables


```python
plt.figure(figsize=(12,7))
corr=data.corr()
sns.heatmap(corr,annot=True,cmap="viridis",vmin=0,vmax=1,fmt='f')
```
