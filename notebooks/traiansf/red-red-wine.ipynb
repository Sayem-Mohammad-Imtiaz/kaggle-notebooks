```python
wine = pd.read_csv('/kaggle/input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
wine.head()
```

```python
X = wine.drop('quality', axis=1)
y = wine['quality']
y.value_counts()
```

```python
y.describe()
```

```python
import seaborn as sns #Visualization

# correlation plot
corr = X.corr()
sns.heatmap(corr, cmap = 'Wistia', annot= True);
```

```python
X.drop('fixed acidity', axis=1, inplace=True)
```

```python
y_normalized = pd.cut(y,2,labels=['bad','good'])
y_normalized.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_normalized, train_size = .66)
X_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

for md in range(1,10):
    dt = DecisionTreeClassifier(max_depth=md)
    scores = cross_val_score(dt, X, y_normalized, cv=10)
    print("max_depth=%d: %0.2f accuracy with a standard deviation of %0.2f" % (md, scores.mean(), scores.std()))
```

```python
classifier = DecisionTreeClassifier(max_depth=4)
model = classifier.fit(X_train, y_train)
classifier.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt # data visualization

plt.figure()
plt.title("Feature importances")
plt.barh(X.columns, model.feature_importances_, 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
model = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt # data visualization

plt.figure()
plt.title("Feature importances")
plt.barh(X.columns, model.feature_importances_, 1)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(solver='liblinear')
scores = cross_val_score(lr, X, y_normalized, cv=10)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
lr.fit(X_train,y_train)
lr.score(X_test,y_test)
```

```python
columns = ['intersect:x_0=1'] + list(X.columns.values)
theta = [lr.intercept_[0]]+list(lr.coef_[0])
parameters = pd.DataFrame({'Columns':columns, 'theta': theta})
parameters
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
scores = cross_val_score(gnb, X, y_normalized, cv=10)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
gnb.fit(X_train, y_train)
gnb.score(X_test, y_test)
```
