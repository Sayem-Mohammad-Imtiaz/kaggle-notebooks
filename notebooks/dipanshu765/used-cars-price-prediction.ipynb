```python
df = pd.read_csv('/kaggle/input/craigslist-carstrucks-data/vehicles.csv')
```

```python
df = df.drop(['county'], axis=1)
```

```python
df.columns
```

```python
df = df.dropna(subset=['size', 'odometer', 'cylinders', 'paint_color', 'lat'])
df = df.drop(['vin', 'condition', 'drive', 'manufacturer'], axis=1)
```

```python
df.isnull().sum()
```

Now let's use KNN to impute the remaining ones.

```python
df.info()
```

```python
df = df.drop(['id', 'url'], axis=1)
```

```python
print(df.nunique())
```

Let's drop non-int columns with more than 15 unique values

```python
cols_to_drop = ['region', 'region_url', 'model', 'image_url', 'description', 'state']
```

```python
df = df.drop(cols_to_drop, axis=1)
```

```python
df2 = df
```

```python
df=pd.get_dummies(data=df, 
               columns=[
                       'cylinders',
                        'fuel', 
                        'title_status', 
                        'transmission', 
                        'size', 
                        'type', 
                        'paint_color'
               ], 
               drop_first=True)
```

```python
imputer = KNNImputer(n_neighbors=2)
df = pd.DataFrame(imputer.fit_transform(df), columns=df.columns)
```
