```python
features = segment_df.iloc[:, [3, 4]].values
print(features.shape)
```

```python
segment_df = pd.get_dummies(segment_df, columns = ['Gender'], prefix = ['Gender'])
segment_df.head()
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    km.fit(features)
    wcss.append(km.inertia_)
wcss
```

```python
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method', fontsize = 20)
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

# k=5

```python
km = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_mean = km.fit_predict(features)
plt.figure(figsize=(10,6))
plt.scatter(features[y_mean == 0, 0], features[y_mean == 0, 1], s = 30, c = 'indigo', label = 'C-1',marker="H")
plt.scatter(features[y_mean == 1, 0], features[y_mean == 1, 1], s = 30, c = 'orange', label = 'C-2',marker="D")
plt.scatter(features[y_mean == 2, 0], features[y_mean == 2, 1], s = 30, c = 'green', label = 'C-3',marker="^")
plt.scatter(features[y_mean == 3, 0], features[y_mean == 3, 1], s = 30, c = 'blue', label = 'C-4',marker="p")
plt.scatter(features[y_mean == 4, 0], features[y_mean == 4, 1], s = 30, c = 'magenta', label = 'C-5',marker="s")
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s = 100, c = 'black' , label = 'Centroid',marker='o')

#plt.style.use('fivethirtyeight')
plt.title('K Means Clustering', fontsize = 20)
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.grid()
plt.show()
```
