**Average Price of the bestelling books through the years**

```python
mean_price = df.groupby('Year').mean('Price')
mean_price
```

```python
mean_price['Year'] = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
```

```python
mean_price
```

**Data Visualization shows that average price for bestselling books is falling over the years**

```python
mean_price.plot(x='Year', y='Price', kind='bar', cmap='Accent', figsize=(11,6))
plt.title('Average price of the bestelling books 2009-2019', fontsize=17)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Average price', fontsize=15)
plt.xticks(rotation='horizontal')
```
