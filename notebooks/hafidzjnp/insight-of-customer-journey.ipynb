```python
#Import EDA
import numpy as np
import pandas as pd

import os

#Import Graph
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Data Preparation
raw_data = pd.read_csv("/kaggle/input/ecommerce-behavior-data-from-multi-category-store/2019-Nov.csv")
```

```python
raw_data.isnull().any(axis=0)
```

```python
raw_data.dropna(axis = 0, inplace = True)
raw_data.isnull().any(axis=0)
```

```python
raw_data.info()
raw_data.head()
```

```python
raw_data['event_time']=pd.to_datetime(raw_data['event_time']).dt.strftime('%w')
```

```python
day =['Monday','Tuesday', 'Wednesday', 'Thursday', 'Friday','Saturday', 'Sunday']
type = raw_data.groupby(['event_time','event_type']).size().reset_index(name='total')
type['event_time'].head()
```

```python
type.head()
```

```python
e_type=raw_data.groupby('event_type').size().reset_index(name='total')
e_type.head(3)

plt.pie(e_type['total'],labels=e_type['event_type'],autopct='%1.1f%%')
plt.title('Customer Behavior')
plt.axis('equal')
plt.show()
```

```python
only_purchases = type.loc[type.event_type == 'purchase'] 
only_cart = type.loc[type.event_type == 'cart']
only_view = type.loc[type.event_type == 'view']
```

```python


plt.bar(only_view['event_time'],only_view['total'],color='green')
plt.title('View a Day', fontsize=14)
plt.xlabel('Day (0-6 = Sunday-Saturday)', fontsize=14)
plt.ylabel('Insight in millions', fontsize=14)
plt.show()
```

```python
only_cart.head(7)

plt.bar(only_cart['event_time'],only_cart['total'],color='blue')
plt.title('Cart a Day', fontsize=14)
plt.xlabel('Day (0-6 = Sunday-Saturday)', fontsize=14)
plt.ylabel('Insight', fontsize=14)
plt.show()
```

```python
only_cart.head(7)

plt.bar(only_purchases['event_time'],only_purchases['total'],color='orange')
plt.title('Purchas a Day', fontsize=14)
plt.xlabel('Day (0-6 = Sunday-Saturday)', fontsize=14)
plt.ylabel('Insight', fontsize=14)
plt.show()
```
