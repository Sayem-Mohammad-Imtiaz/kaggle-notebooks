```python
#import EDA
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

#Import Graph
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv('/kaggle/input/ecommerce-behavior-data-from-multi-category-store/2019-Nov.csv')
```

```python
raw_data.dropna(axis=0,inplace=True)
raw_data.isnull().any(axis=0)
```

```python
raw_data['event_time']=pd.to_datetime(raw_data['event_time']).dt.strftime('%A')
raw_data.head()
```

```python
only_purchases = raw_data.loc[raw_data.event_type == 'purchase'] 
only_cart = raw_data.loc[raw_data.event_type == 'cart']
only_view = raw_data.loc[raw_data.event_type == 'view']
```

```python
#TOP CART
cart = only_cart.groupby(['brand','category_code','event_type']).size().reset_index(name = 'total').sort_values(by='total', ascending=False).nlargest(10,'total')
cart
```

```python
#TOP VIEW
view = only_view.groupby(['brand','category_code','event_type']).size().reset_index(name = 'total').sort_values(by='total', ascending=False).nlargest(10,'total')
view
```

```python
#TOP PURCHASE
purchase = only_purchases.groupby(['brand','category_code','event_type']).size().reset_index(name = 'total').sort_values(by='total', ascending=False).nlargest(10,'total')
purchase
```

```python
#top purchase
data=only_purchases.groupby(by='user_session').agg(Date_order=('event_time',lambda x: x.max()),
                                                  user_id=('user_id',lambda x: x.unique()),
                                          Quantity=('user_session','count'),
                                         money_spent=('price','sum')).reset_index(drop=True).sort_values(by='Quantity', ascending=False).nlargest(5,'Quantity')
data
```
