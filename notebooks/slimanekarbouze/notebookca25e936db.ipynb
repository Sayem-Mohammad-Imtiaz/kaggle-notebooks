```python
#Structure de données : DataFrame
import pandas as pd
import numpy as np


df = pd.read_csv('../input/housesalesprediction/kc_house_data.csv')
df['log_sqft_lot'] = np.log(1+df['sqft_lot'])
df['log_price'] = np.log(1+df['price'])


df1 = df[ ['log_sqft_lot', 'floors', 'log_price'] ]

#80% des données apprentissage
#20% des données pour le test
train_df = df1[0:17000]
test_df  = df1[17000:]

# crée dummies variables
dum = pd.get_dummies(df.zipcode)

merged = pd.concat([df1,dum],axis='columns')

# supprimer la colonne log_sqft_lot
fin = merged.drop(['log_sqft_lot'], axis='columns')

# supprimer la colonne 98199
fin1 = fin.drop([98199], axis='columns')

# supprimer la colonne log_price
fin2 = fin1.drop(['log_price'], axis='columns')

#80% des données apprentissage
#20% des données pour le test
train_df = merged[0:17000]
test_df  = merged[17000:]

#Separer les données en X et Y 
train_X = train_df[ ['log_sqft_lot', 'floors'] ].values
train_Y = train_df[ 'log_price' ].values

test_X = test_df[ ['log_sqft_lot', 'floors'] ].values
test_Y = test_df[ 'log_price' ].values

print( train_X.shape, test_X.shape )
print( train_Y.shape, test_Y.shape )

from sklearn.linear_model import LinearRegression
import keras
from keras import layers
from keras.models import Model
from keras.optimizers import SGD
model = LinearRegression()

inpt = layers.Input( shape=(train_X.shape[1],) )
x    = layers.Dense(2) ( inpt )
actv = layers.Activation('relu') (x)
output = layers.Dense(1) (actv)

model = Model(inpt,output)
model.summary()
```

```python
opt = SGD() #choisir le gradient descent comme optimizer pour la mise à jour des poids
#mse : mean squared error

model.compile( optimizer=opt, loss='mse' )
model.fit( train_X, train_Y, epochs=5 )
```
