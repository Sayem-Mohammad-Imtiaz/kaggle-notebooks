```python
X = A[['YearsExperience']]
Y = A[['Salary']]
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=.25,random_state=10)
sb.distplot(X)
sb.distplot(xtrain)
```

```python
from sklearn.linear_model import LinearRegression
slr = LinearRegression()
model = slr.fit(xtrain,ytrain)
pred = model.predict(xtest)
pred
```

```python
test1 = xtest[['YearsExperience']]
test1['actual_value'] = ytest
test1['predicted_value'] = pred
test1['difference error'] = ytest-pred
test1
```

```python
import sklearn.metrics as m
print("Mean Absolute Error is : {}".format(m.mean_absolute_error(ytest,pred)))
print("Mean Absolute Error is : {}".format(m.mean_squared_error(ytest,pred)))
print("Predicted value is {}".format(model.predict([[7.3]])[0][0]))
b0 = model.intercept_
b1 = model.coef_
print("Predicted value is {}".format((b0+b1*7.3)[0][0]))
print("r2_score is {}".format(m.r2_score(ytest,pred)))
```

```python
#Visualization
fig, axis = plt.subplots(1, 2, figsize=(15, 5))
axis[0].set_title('Training Data', fontsize=15)
axis[0].scatter(xtrain, ytrain, s=75, color='red')
axis[0].plot(xtrain, b0 + b1*xtrain)
axis[1].set_title('Testing Data', fontsize=15)
axis[1].scatter(xtest, ytest, s=75, color='red')
axis[1].plot(xtest, b0 + b1*xtest)
```
