```python
# Directive pour afficher les graphiques dans Jupyter
%matplotlib inline
# Pandas : librairie de manipulation de données
# NumPy : librairie de calcul scientifique
# MatPlotLib : librairie de visualisation et graphiques
# SeaBorn : librairie de graphiques avancés
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, roc_auc_score,auc, accuracy_score
from matplotlib import pyplot as plt
from sklearn import preprocessing
import seaborn as sns
from sklearn import model_selection
from sklearn import ensemble
import xgboost as XGB
from sklearn.model_selection import validation_curve
t = pd.read_csv("../input/pima-indians-diabetes-database/diabetes.csv")
def replace_0(df,col) :
    df1 = df.copy()
    n = df.shape[0]
    m = df[col].mean()
    s = df[col].std()
    for i in range(n) :
        if df.loc[i,col]==0 :
            df1.loc[i,col] = np.random.normal(m,s)
    return df1
```

```python
t.head().T
```

# Affichage du dataset selon chaque variable

```python
t = replace_0(t,'Pregnancies')
t = replace_0(t,'Glucose')
t = replace_0(t,'BloodPressure')
t = replace_0(t,'SkinThickness')
t = replace_0(t,'Insulin')
t = replace_0(t,'BMI')
t = replace_0(t,'DiabetesPedigreeFunction')
t = replace_0(t,'Age')
fig = sns.FacetGrid(t, hue="Outcome", aspect=3, palette="Set2")
fig.map(sns.kdeplot, "Pregnancies", shade=True)
fig.add_legend()
fig = sns.FacetGrid(t, hue="Outcome", aspect=3, palette="Set2")
fig.map(sns.kdeplot, "Glucose", shade=True)
fig.add_legend()
fig = sns.FacetGrid(t, hue="Outcome", aspect=3, palette="Set2")
fig.map(sns.kdeplot, "BloodPressure", shade=True)
fig.add_legend()
fig = sns.FacetGrid(t, hue="Outcome", aspect=3, palette="Set2")
fig.map(sns.kdeplot, "SkinThickness", shade=True)
fig.add_legend()
fig = sns.FacetGrid(t, hue="Outcome", aspect=3, palette="Set2")
fig.map(sns.kdeplot, "Insulin", shade=True)
fig.add_legend()
fig = sns.FacetGrid(t, hue="Outcome", aspect=3, palette="Set2")
fig.map(sns.kdeplot, "BMI", shade=True)
fig.add_legend()
fig = sns.FacetGrid(t, hue="Outcome", aspect=3, palette="Set2")
fig.map(sns.kdeplot, "DiabetesPedigreeFunction", shade=True)
fig.add_legend()
fig = sns.FacetGrid(t, hue="Outcome", aspect=3, palette="Set2")
fig.map(sns.kdeplot, "Age", shade=True)
fig.add_legend()
```

**Formatage des données**

```python
dataset = t[['Glucose','Outcome']]
dataset.count()
```

```python
dataset = t.drop(['BloodPressure', 'Insulin', 'SkinThickness'], axis=1)
X = dataset.drop(['Outcome'], axis=1)
y = dataset.Outcome
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print(X_train.shape)
print(X_test.shape)
```

# Méthode 1 : Logistic Regression

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_lr = lr.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_lr))
```

```python
print(accuracy_score(y_test,y_lr))
```

```python
print(classification_report(y_test, y_lr))
```

```python
probabilities = lr.predict_proba(X_test)
print(probabilities)
```

```python
dfProbabilities = pd.DataFrame(probabilities,columns=['diabetic','not diabetic'])
dfProbabilities['truth'] = np.array(y_test)
```

```python
dfProbabilities
```

# Méthode 2 : Forêts Aléatoires

```python
param_grid = {
              'n_estimators': [10, 100, 500],
              'min_samples_leaf': [1, 20, 50]
             }
estimator = ensemble.RandomForestClassifier()
gridSearch = model_selection.GridSearchCV(estimator, param_grid)
```

```python
gridSearch.fit(X_train, y_train)
```

```python
print(gridSearch.best_params_)
```

```python
rf2 = gridSearch.best_estimator_
y_rf2 = rf2.predict(X_test)
print(classification_report(y_test, y_rf2))
```

# Méthode 3: XGBOOST

```python
xgb  = XGB.XGBClassifier()
xgb.fit(X_train, y_train)
y_xgb = xgb.predict(X_test)
cm = confusion_matrix(y_test, y_xgb)
print(cm)
print(classification_report(y_test, y_xgb))
```
