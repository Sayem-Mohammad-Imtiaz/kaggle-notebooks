```python
df = pd.read_csv("../input/bostonhoustingmlnd/housing.csv")
df.head(10).T
```

# On cherche une corrélation et on affiche via une HEATMAP :)

```python
tabcorr = df.corr()
plt.figure(figsize=(12,12))
sns.heatmap(abs(tabcorr), cmap="coolwarm")
```

# Il y a peu de paramètres, mais on se rend compte que le PTRATIO n'est pas très pertinent.

```python
sns.clustermap(abs(tabcorr), cmap="coolwarm")
```

```python
correlations = tabcorr.MEDV
correlations = correlations.drop(['MEDV'],axis=0)
print(correlations)
```

# On remarque que RM et LSTAT sont très pertinents, alors que PTRATIO est quasiment inutile.

```python
df = df.drop(['PTRATIO'], axis=1)
X = df.drop(['MEDV'], axis=1)
y = df.MEDV
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
```

# Une fois que l'on a choisi les données, on utilise le dataset avec les méthodes suivantes :

# LINEAR REGRESSION

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
y_pred = lm.predict(X_test)
```

```python
plt.figure(figsize=(12,12))
plt.scatter(y_test, y_pred)
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()], color='red', linewidth=3)
plt.xlabel("MEDV")
plt.ylabel("Prediction des MEDV")
plt.title("Realité vs prédiction")
```

```python
sns.distplot(y_test-y_pred)
```

```python
print(np.sqrt(mean_squared_error(y_test, y_pred)))
scoreR2 = r2_score(y_test, y_pred)
print(scoreR2)
lm.score(X_test,y_test)
```

# Régression par forêt aléatoire :)

```python
X = df.drop(['MEDV'], axis=1)
y = df.MEDV
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=1)
from sklearn import ensemble
rf = ensemble.RandomForestRegressor()
rf.fit(X_train, y_train)
y_rf = rf.predict(X_test)
print(rf.score(X_test,y_test))
```

```python
plt.figure(figsize=(12,12))
plt.scatter(y_test, y_rf)
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()], color='red', linewidth=3)
plt.xlabel("MEDV")
plt.ylabel("Prediction des MEDV")
plt.title("Realité vs prédiction")
```

```python
sns.distplot(y_test-y_rf)
```

# XGBOOST

```python
import xgboost as XGB
xgb  = XGB.XGBRegressor()
xgb.fit(X_train, y_train)
y_xgb = xgb.predict(X_test)
print(xgb.score(X_test,y_test))

plt.figure(figsize=(12,12))
plt.scatter(y_test, y_xgb)
plt.plot([y_test.min(),y_test.max()],[y_test.min(),y_test.max()], color='red', linewidth=3)
plt.xlabel("MEDV")
plt.ylabel("Prediction des MEDV")
plt.title("Realité vs prédiction")
```
