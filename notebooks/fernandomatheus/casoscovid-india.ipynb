# Análise Exploratória e Seleção

```python
dados = pd.read_csv('../input/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv')
dados.head().style.set_properties(**{'background-color':'black','color': 'white'})
```

```python
dados.describe().style.set_properties(**{'background-color':'black','color': 'white'})
```

```python
dados.value_counts()
```

```python
dados1=dados.sort_values(by="Deaths", ascending=False)
dados2=dados.sort_values(by="Total Cases", ascending=False)
CasosTotal=dados2.iloc[0:5]
CasosMorte= dados1.iloc[0:5]
```

# As 6 cidades com maiores casos de mortos na India

```python
sns.catplot(x="State/UTs", y="Deaths", kind="bar", data=CasosMorte,aspect=3)
```

# Total de casos, casos ativos e altas

```python
sns.catplot(x="State/UTs", y="Discharged", kind="bar", data=CasosTotal,aspect=3)
#Altas
print("Número de Casos Total nas cidades exibidas:",CasosTotal["Total Cases"].sum())
```

```python
dados3=dados.sort_values(by="Active", ascending=False)
CasosAtivos=dados3.iloc[0:5]
sns.catplot(x="State/UTs", y="Active", kind="bar", data=CasosAtivos,aspect=3)
CasosAtivos
```

# A quantidade de casos recuperados/ativos/totais
### Total de casos ativos de COVID é inferior a 1%, que mostra um controle da epidemia no país como um todo

```python
listd4 = [378181, 31993644, 32810845]
explode = (0.1, 0.1, 0.1) #explode Recuperados
labels = ['Ativos','Recuperados','Casos Totais']
fig1, ax1 = plt.subplots()
ax1.pie(listd4, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True)
ax1.axis('equal')
plt.show()
```
