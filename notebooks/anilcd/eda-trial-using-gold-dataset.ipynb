```python
df = pd.read_csv("/kaggle/input/gold-prices/monthly_csv.csv")
df.head()
```

```python
df.plot(kind = "line", x = "Date", y = "Price")
plt.show()
```

```python
after_2000 = df[df["Date"] > "2000"]
after_2000.head()
```

```python
after_2000.plot(kind = "line", x = "Date", y = "Price")
```

```python
after_2000["Change"] = after_2000["Price"].diff()
after_2000["G/L"] = np.where(after_2000["Change"] > 0, 'G', 'L')
after_2000["%Change"] = (after_2000["Price"].pct_change()) * 100
after_2000.head()
```

```python
after_2000.plot(kind = "line", x = "Date", y = "Change")
plt.show()
```

```python
after_2000.plot(kind = "scatter", x = "Date", y = "Change")
plt.show()
```

```python
after_2000.plot(kind = "line", x = "Date", y = "%Change")
plt.show()
```

```python
after_2000.plot(kind = "scatter", x = "Date", y = "%Change")
plt.show()
```

```python
after_2000.plot(kind = "hist", y = "%Change", bins = 100)
plt.show()
```

```python
gain, loss = after_2000["G/L"].value_counts(dropna=False)
gain_perc = gain / (gain + loss)
print("Monthly gold increment probability", gain_perc)
```

```python
after_2000.boxplot(column="%Change",by = "G/L")
```

```python
after_2000["Date"] = pd.to_datetime(after_2000["Date"])
after_2000 = after_2000.set_index("Date")
after_2000.head()
```

```python
after_2000_avg = after_2000.resample("A").mean()
after_2000_avg
```

```python
after_2000_avg.plot(kind = "hist", y = "%Change", bins = 100)
plt.show()
```
