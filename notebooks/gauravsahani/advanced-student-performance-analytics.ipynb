```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
df.head()
```

```python
df['average score'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)
```

```python
df.head()
```

Lets plot a quick histogram!

```python
df.hist()
```

Execute the below code, in order to have a view for the histogram of the Average scores od students!

```python
plt.hist(df['average score'], color = 'blue', edgecolor = 'black',bins=10)

# Add labels
plt.title('Combined Average Scores')
```

Now, We'll be executing below code to find the degree comparision based on the parental level of education data we have, with respect to the gender!

```python
sns.set_style('whitegrid')
sns.set(rc={'figure.figsize':(10.7,9.7)})
sns.countplot(x = 'parental level of education', data = df, hue='gender')
plt.title('Degree Comparison')
locs, labels = plt.xticks()
plt.setp(labels, rotation=20)
```

```python
sns.catplot(x="race/ethnicity", y="average score", data=df,kind="swarm",hue='gender',height=6,aspect=3)
```

```python
plt.figure(figsize=(15,6))
plt.subplot(1, 4, 1)
sns.barplot(x='test preparation course',y='math score',data=df,hue='lunch',palette='Paired')
plt.title('Math Score')
plt.subplot(1, 4, 2)
sns.barplot(x='test preparation course',y='reading score',data=df,hue='lunch',palette='Paired')
plt.title('Reading Scores')
plt.subplot(1, 4, 3)
sns.barplot(x='test preparation course',y='writing score',data=df,hue='lunch',palette='Paired')
plt.title('Writing Scores')
plt.subplot(1, 4,  4)
sns.barplot(x='test preparation course',y='average score',data=df,hue='lunch',palette='Paired')
plt.title('Total Score')
plt.show()
```

```python
plt.figure(figsize=(30,20))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9,
                      wspace=0.5, hspace=0.2)

plt.subplot(142)
plt.title('Ethinicity',fontsize = 20)
df['race/ethnicity'].value_counts().plot.pie(autopct="%1.1f%%")
```
