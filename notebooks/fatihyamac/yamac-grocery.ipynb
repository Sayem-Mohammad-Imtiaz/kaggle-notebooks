```python
# J'ai transformé les données à la forme de liste;
data = list(df["products"].apply(lambda x : x.split(',')))
data
```

```python
# J'ai choisi l'algorithme Apriori (qui connait les données true/false) pour cetype de préparation; 
# J'ai importé l'algoritme Apriori de bibliothèque mlextend avec la bibliothèque mlextend;
# J'ai transformé les données à la format true/false; 
!pip install mlxtend
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data,columns=te.columns_)
df.head()
```

```python
# J'ai donné la valeur minimum d'algoritme comme 0.2 (j'ai eliminé les données sous 0.2);
# Et l'argument verbose comme 1 (afin de savoir le nombre de combination ); 
from mlxtend.frequent_patterns import apriori
freq_items = apriori(df,min_support=0.20,use_colnames = True, verbose = 1)
freq_items.head()
```
