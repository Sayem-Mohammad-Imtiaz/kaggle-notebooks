# **Analyze Phase**

```python
plt.figure(figsize=(10,8))
merged_corr = Activity.corr()
sns.heatmap(data = merged_corr,annot=True)
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Day', data=Activity, order=["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"])
plt.xticks(rotation=90)
plt.title("Number of users per day")
```

> **Number of users per day**

Above this plot represent the number of users per day. Most of the users applied the smart devices on **Wednesday,Tuesday, and Thrusday**.

```python
plt.figure(figsize=(15,10))
sns.catplot(data=Activity,kind='bar', x='Day',y='TotalSteps',order=["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"])
plt.xticks(rotation=90)
plt.title("Number of step count per day")
```

> **Number of step count per day**

From this day plot, here have some interesting findings. It shows the step count of the **weekend day is higher than the weekday**. It can be possible reasons is that on a weekday most of the people work on sitting.

```python
plt.figure(figsize=(10,8))
sns.catplot(data=Activity,kind='bar', x='Day',y='Calories',order=["Sunday","Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"])
plt.xticks(rotation=90)
plt.title("Calories burned per day")
```

> **Calories burned per day**

The interesting part of this chart is shows the burned calories **quantity very closely of every day**. One possible explanation week day they have spend time on their office and weekend spend on their home.

```python
plt.figure(figsize=(10,8))
sns.catplot(data=Activity,kind='bar', x='Month',y='Calories')
plt.xticks()
plt.title("Calories burned per month")
```

> **Calories burned per month**

Here we see the dataset has a limitation. It contains the data only in **April and May**. So we can not show which month people are most active of their calories burned.

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=Activity, x='TotalSteps',y='Calories',hue='TotalDistance')
plt.xticks(rotation=90)
plt.title("Calories burned in TotalStep") 
```

> **Calories burned in TotalStep**

There have **linear relation** between the calories and TotalStep. Most of the users have within **15000** step count. 

```python
Activity['TotalMinutes'] = Activity['LightlyActiveMinutes'] + Activity['FairlyActiveMinutes'] + Activity['VeryActiveMinutes'] + Activity['SedentaryMinutes']

activitiesm_means = Activity['TotalMinutes'].mean()
lightlym_pct = (Activity['LightlyActiveMinutes'].mean()/activitiesm_means) * 100
fairlym_pct = (Activity['FairlyActiveMinutes'].mean()/activitiesm_means) * 100
verym_pct = (Activity['VeryActiveMinutes'].mean()/activitiesm_means) * 100
sedentarym_pct = (Activity['SedentaryMinutes'].mean()/activitiesm_means) * 100


plt.figure(figsize=(15,8))
plt.pie([ fairlym_pct, verym_pct,lightlym_pct,sedentarym_pct], 
        labels = [ "Fairly Active Minutes", "Very Active Minutes", "Light Active Minutes", "Sedentary Minutes"], 
        colors = ['m', 'g', 'y', 'c'], 
        wedgeprops = {"edgecolor": "black"}, 
        explode = [0, 0, 0, 0.1], 
        autopct = "%1.1f%%")
plt.title("Percentage of Activity in Minutes")
```

> **Percentage of Activity in Minutes**

According to this chart, **81.3%** of users spend is sedentary minutes, and only **15.8%** of users spend light active minutes. One probable explanation is that people use smart gadgets all day, not just when exercising.

```python
activities_means = Activity["TotalDistance"].mean()

lightly_pct = (Activity['LightActiveDistance'].mean()/activities_means) * 100
moderately_pct = (Activity['ModeratelyActiveDistance'].mean()/activities_means) * 100
very_pct = (Activity['VeryActiveDistance'].mean()/activities_means) * 100

plt.figure(figsize=(10,8))
plt.pie([lightly_pct, moderately_pct, very_pct], 
        labels = ["Light Active Distance", "Moderately Active Distance", "Very Active Distance"], 
        colors = ['#9d8189', '#ffe5d9', '#d8e2dc'],  
        explode = [0.1, 0, 0], 
        autopct = "%1.1f%%")
plt.title("Percentage of Activity in Distance")
plt.show()
```

```python
sleeping_means = Sleep["TotalTimeInBed"].mean()

sleep_pct = (Sleep['TotalMinutesAsleep'].mean()/sleeping_means) * 100
awake_pct = (Sleep['TotalMinutesAwake'].mean()/sleeping_means) * 100

plt.figure(figsize=(10,8))
plt.pie([sleep_pct, awake_pct], 
        labels = ['TotalMinutesAsleep','TotalMinutesAwake'], 
        colors = ['b', 'y'],  
        explode = [0.1, 0], 
        autopct = "%1.1f%%")
plt.title("Percentage of Total_Asleep vs Total_Awake ")
plt.show()
```

> **Percentage of Total_Asleep vs Total_Awake**

91.5% of users fall asleep after getting the bed.
