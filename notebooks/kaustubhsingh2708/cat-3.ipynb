Q.9)Is Any supervised machine learning possible ? if yes explain.

Ans.)The given dataset shows us various details of countries over past 5 years. Supervised machine learning is the task of learning a function that maps an input to an output based on example input-output pairs i.e it infers a function from labeled training data consisting of a set of training examples. The algorithm learns on a labeled dataset, providing an answer key that the algorithm can use to evaluate its accuracy on training data.The given dataset has enough data for machine to train and predict a data for the next year with a high accuracy and hence we can say that supervised machine learning is possible for the given data set.
