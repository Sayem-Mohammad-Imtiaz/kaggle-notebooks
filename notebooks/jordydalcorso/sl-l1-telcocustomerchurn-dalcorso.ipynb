```python
#New dataset from the existing one with less attributes (not sure if this is a new copy or a pointer)
data1 = data[['gender', 'Partner', 'MonthlyCharges', 'Churn']]
```

```python
#This is how the new dataset looks like
data1.head(5)
```

```python
#gender: Male=0, Female=1 • Partner: No=0, Yes=1 • Churn: No=0, Yes=1
#Converting binary categories to numerical binary values

data1['gender'] = data1['gender'].map( {'Male': 0, 'Female': 1} ).astype(int)
data1['Partner'] = data1['Partner'].map( {'No': 0, 'Yes': 1} ).astype(int)
data1['Churn'] = data1['Churn'].map( {'No': 0, 'Yes': 1} ).astype(int)

#This is how the ds looks like after the process
data1.head(5)
```

```python
#The method (I don't think this is a method) iloc extract only the fourth column of the ds
#The method drop() delete the column from the dataset. Using inplace=True we avoid creating a new ds and we operate on the existing one

data1Churn = data1.iloc[:,3]
data1.drop('Churn', inplace=True, axis=1)

#This command shows us we're deaing with a Series object
type(data1Churn)
```

```python
#Finally we create a LogisticRegression object and we fit it with X=data1 and y=data1Churn
#At the end the coefficient associated to the independent variables are printed

logreg = LogisticRegression()
logreg.fit(data1, data1Churn)
logreg.coef_[0]
```
