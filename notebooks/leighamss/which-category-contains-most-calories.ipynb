```python
import numpy as np 
import seaborn as sns 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = 11,8
```

```python
mc= pd.read_csv('../input/menu.csv')
#Clean Data
mc.isnull().any()
```

<h1> Items on McDonald’s Menu with the most Calories(By Category) </h1>
<p>I honestly didn't expect close to 2000 Calories from one sandwich, wow ! </p>

```python
Most_Calories = mc.groupby('Category').max().sort_values('Calories',ascending=False)
sns.swarmplot(data =Most_Calories, x= Most_Calories.index,y = 'Calories',hue ='Item',size =10 )
plt.tight_layout()
```

<h1> Items on McDonald’s Menu with the least Calories(By Category) </h1>
<p>Whether it's looking at the item within the category with the least amount of calories or the item with the most,
both times the beef and pork category contain the most categories.</p>
<li>Please note that items with less than 5 calories can be listed as having zero calories according to the FDA <a href="https://www.accessdata.fda.gov/scripts/cdrh/cfdocs/cfcfr/CFRSearch.cfm?fr=101.60">Click Here for more information</a>  </li>

```python
Least_Calories = mc.groupby('Category').min().sort_values('Calories',ascending=False)
sns.swarmplot(data =Least_Calories, x= Least_Calories.index,y = 'Calories',hue ='Item',size =10 )
plt.tight_layout()
```

<h3>McDonald's mostly sells coffee and tea and sells a small amount of Salads in comparison  </h3>
<p> I had no idea that the mass majority of the items on McDonald's menu were coffee and Tea, But I'm not surprised the item they sell the least of is Salads 
<li>36 % of McDonald s menu is comprised of Coffee and Tea while only 2 % is comprised of salads</li></p>

```python
amount = mc.groupby('Category').count()
amount_count = amount[['Item']].sort_values('Item',ascending= False)
amount_count
```

<h1>Most Fiber </h1>
<p>Not only are salads healthy but also contain the most fiber out of the other items on the McDonald’s menu.</p>

```python
sns.boxplot(data= mc, x = 'Category',y = 'Dietary Fiber')
plt.tight_layout
plt.show()
```
