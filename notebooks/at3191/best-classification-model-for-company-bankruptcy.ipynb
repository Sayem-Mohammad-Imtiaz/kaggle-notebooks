```python
import pandas as pd
```

```python
company = pd.read_csv("../input/company-bankruptcy-prediction/data.csv")
```

```python
company
```

```python
company.describe()
```

```python
len(list(company))
```

```python
train_x = company.iloc[0:5000,1:]
train_y = company.iloc[0:5000,0:1]
test_x = company.iloc[5000:6818,1:]
test_y = company.iloc[5000:6818,0:1]
print(len(train_x),len(train_y),len(test_x),len(test_y))
```

```python
len(list(train_x))
```

```python
#KNN

from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=3)
model.fit(train_x, train_y)

testpredict = model.predict(test_x)
trainpredict = model.predict(train_x)


print("test_score",model.score(test_x, test_y))
print("train_score",model.score(train_x, train_y))

print(trainpredict)
```

```python
##Logistic Regression

from sklearn.linear_model import LogisticRegression
model = LogisticRegression(random_state=0).fit(test_x, test_y)


print("test_score",model.score(test_x, test_y))
print("train_score",model.score(train_x, train_y))
```

```python
## DecisionTree

from sklearn import tree
model = tree.DecisionTreeClassifier()
model = model.fit(test_x, test_y)
print("test_score",model.score(test_x, test_y))
print("train_score",model.score(train_x, train_y))
```

```python
##Random Forrest

from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification

model = RandomForestClassifier(max_depth=2, random_state=0)
model.fit(train_x, train_y)

print("test_score",model.score(test_x, test_y))
print("train_score",model.score(train_x, train_y))

```

```python
#SVM
from sklearn import svm

model = svm.SVC()
model.fit(test_x, test_y)

print("test_score",model.score(test_x, test_y))
print("train_score",model.score(train_x, train_y))
```

```python
#Bagging
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
from sklearn.datasets import make_classification

model = BaggingClassifier(base_estimator=SVC(),
                        n_estimators=10, random_state=0).fit(train_x, train_y)

print("test_score",model.score(test_x, test_y))
print("train_score",model.score(train_x, train_y))
```

```python
#Boosting 
from sklearn.ensemble import GradientBoostingClassifier
model = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,
    max_depth=1, random_state=0).fit(train_x, train_y)

print("test_score",model.score(test_x, test_y))
print("train_score",model.score(train_x, train_y))
```

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(solver='lbfgs', alpha=1e-5,
                    hidden_layer_sizes=(5, 2), random_state=1)

model.fit(train_x, train_y)

print("test_score",model.score(test_x, test_y))
print("train_score",model.score(train_x, train_y))

```

*Boosting is the best model to predict*
