# DATA ANALYSIS AND VISUALIZATION

ANALYSIS OF COUNTRY INDIA

```python
df_India=df[df["iso_code"]=="IND"].copy()
df_India.drop(df_India.index[df_India['people_vaccinated'] == 0], inplace = True)
df_India.head()
```

Lets see whether the daily_vaccination is equal or not to the daily_vacinations_raw. For these we will perform a hypithesis testing using t statistics where H0:u1=u2 and Ha:u1!=u2 . And as it is two-tailed t test Level of Significance is 0.05/2=0.025 . n is 76(39+39-2)

```python
df_India.shape
```

```python
from scipy import stats
from scipy.stats import ttest_ind
print(ttest_ind(df_India["daily_vaccinations"],df_India["daily_vaccinations_raw"],equal_var=False))
print(stats.t.ppf(.025,76))
```

As L.O.S is less then p value we accept null hypothesis . That is daily_vaccination is equal to the daily_vaccinations_raw

```python
plt.figure(figsize=(20,10))
sns.lineplot(data=df_India,x="date",y="daily_vaccinations",marker='d',markersize=20,color='k')
plt.title("Vaccination line of India")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data=df_India, x="date",y="total_vaccinations",hue="vaccines")
plt.xticks(rotation=90)
plt.title("Vaccination line of India")
plt.show()
```

# WHICH COUNTRY HAD GIVEN PEOPLE THE FASTEST

This can be seen if we see the people vaccinated per hundred

```python
cols = ['total_vaccinations', 'iso_code', 'vaccines','total_vaccinations_per_hundred']
vac_amt = df.groupby('country')[cols].max().sort_values('total_vaccinations_per_hundred',ascending=False).dropna(subset=["total_vaccinations"])
vac_amt=vac_amt.iloc[:10]
plt.figure(figsize=(20,10))
sns.barplot(data=vac_amt,x="total_vaccinations_per_hundred",y=vac_amt.index)
plt.title("Top ten country with vaccination per houndred")
plt.show()
```

# WHICH COUNTRY HAS THE MOST VACCINATED PEOPLE

```python
cols=["iso_code","total_vaccinations","vaccines"]
vac_plp=df.groupby("country")[cols].max().sort_values('total_vaccinations',ascending=False)
plt.figure(figsize=(20,10))
sns.barplot(data=vac_plp,x="total_vaccinations",y=vac_plp.index)
plt.show()
```

# WHICH VACCINE IS USED THE MOST

```python
cols = ['total_vaccinations', 'iso_code', 'vaccines','total_vaccinations_per_hundred']
vac_amt1 = df.groupby('vaccines')[cols].max().sort_values('total_vaccinations',ascending=False)
plt.figure(figsize=(20,10))
sns.barplot(data=vac_amt1,x=vac_amt1.index,y="total_vaccinations")
plt.xticks(rotation=90)
plt.show()
```

# TOP TEN COUNTRIES WITH VACCINATED PEOPLE PER MILLION

```python
cols=['total_vaccinations', 'iso_code', 'vaccines','daily_vaccinations_per_million']
amt_mlm=df.groupby('country')[cols].max().sort_values('daily_vaccinations_per_million',ascending=False)
amt_mlm=amt_mlm.iloc[:10]
plt.figure(figsize=(10,15))
sns.barplot(data=amt_mlm,x="daily_vaccinations_per_million",y=amt_mlm.index)
plt.title("country with people Vaccinated per million")
plt.show()
```
