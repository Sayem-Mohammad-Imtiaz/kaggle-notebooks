```python
y = pd.DataFrame(DF['Attrition_Flag']) # Attrition_Flag is the dependent variable on which we are going to preditction.
```

```python
DFx = DF.drop(['Attrition_Flag', 'CLIENTNUM'], axis =1)
```

We will convert all catogorical attributes in to dummy variable as below.

```python
edu_dummy = pd.get_dummies(DFx['Education_Level'])
Marital_Status_dummy = pd.get_dummies(DFx['Marital_Status'])
Income_dummy =  pd.get_dummies(DFx['Income_Category'])
Card_dummy =  pd.get_dummies(DFx['Card_Category'])
```

```python
DFx = DFx.drop (['Education_Level'], axis = 1)
```

```python
DFx = DFx.drop (['Marital_Status'], axis = 1)
DFx = DFx.drop (['Income_Category'], axis = 1)
DFx = DFx.drop (['Card_Category'], axis = 1)
```

```python
edu_dummy.head()
```

```python
DFx = pd.concat([edu_dummy,Marital_Status_dummy,Income_dummy,Card_dummy], axis = 1)
```

```python
x = pd.DataFrame(DFx)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```
