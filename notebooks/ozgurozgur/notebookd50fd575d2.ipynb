```python
# data = pd.read_csv('C:\\Users\\kbc\Downloads\weatherAUS.csv')
# df = data.copy()
# df.info()
# df.nunique()
data = pd.read_csv('../input/weather-dataset-rattle-package/weatherAUS.csv')
df = data.copy()
df.head()
```

```python
df = df.dropna(subset=['RainTomorrow', 'RainToday'])
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y/%m/%d')

df['Year'] = df['Date'].dt.year.astype('int16')
df['Month'] = df['Date'].dt.month.astype('int16')
df['YearMonth']=df['Date'].dt.strftime('%Y-%m')

df.head()
```

```python
# Frequency of categorical variables with histogram


cats=['Location','WindGustDir','WindGustSpeed','WindDir9am', 'WindDir3pm', 'RainToday', 'RainTomorrow']
for i in cats:
    fig = px.histogram(df, x=i)
    fig.show()
```

```python
# Distribution of numerical variables with histogram and boxplot/violin
# Additionally RainToday and RainTomorrow variables as color dimension

cols=list(df.columns)

for i in cats:
    cols.remove(i)
nums=cols

for i in nums:
    fig = px.violin(df, x=i, color="RainTomorrow")
    fig.show()
    
    fig= px.histogram(data_frame=df, x=i, color="RainTomorrow")
    fig.show()
```

```python
# Scatter matrix and/or OLS Scatter Plot for numerical variables
# Additionally RainToday and RainTomorrow variables as color dimension

fig = px.scatter_matrix(df, dimensions=['MinTemp','MaxTemp'],   color="RainToday") # remove underscore

fig.update_traces(diagonal_visible=False)
fig.show()
```

```python
# Time series (preferably monthly) line chart for numerical variables
# RainToday and RainTomorrow variables as 3rd variable with color dimension
# Tip: Use groupby method to group by month/year and calculate mean values

df2 = df.groupby(['YearMonth','RainToday']).mean().reset_index()

fig = px.line(df2, x="YearMonth", y="MaxTemp", color='RainToday')
fig.show()
# df2
```
