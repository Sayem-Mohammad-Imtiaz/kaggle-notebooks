### **Importing of datasets**

```python
df=pd.read_csv('/kaggle/input/churn-modelling/Churn_Modelling.csv')
df.head()
```

### **Splitting**

```python
X=df.iloc[:,3:13]
y=df.iloc[:,13]
```

### **Creating of dummy variables**

```python
geography=pd.get_dummies(X['Geography'],drop_first=True)
gender=pd.get_dummies(X['Gender'],drop_first=True)
```

### **Dropping of unnecessary features**

```python
X=X.drop(['Geography','Gender'],axis=1)
X.head()
```

### **Concating of dummy features**

```python
X=pd.concat([X,geography,gender],axis=1)
X.head()
```

### **Creating Train-Test Dataset**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

### **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

### **Importing of Keras libraries**

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LeakyReLU,PReLU,ELU
from keras.layers import Dropout
```

### **Model Creation**

```python
classifier=Sequential()
classifier.add(Dense(units=10,kernel_initializer='he_uniform',activation='relu',input_dim=11))
classifier.add(Dropout(0.3))
classifier.add(Dense(units=10,kernel_initializer='he_uniform',activation='relu'))
classifier.add(Dropout(0.4))
classifier.add(Dense(units=10,kernel_initializer='he_uniform',activation='relu'))
classifier.add(Dropout(0.3))
classifier.add(Dense(units=1,kernel_initializer='glorot_uniform',activation='sigmoid'))
# classifier.complier(optimizer='adam',loss=binary_)
```

```python
classifier.summary()
```

Formula of para=no of neurons in previous layer X no of neurons in current layer +1 X no of neurons in current layer

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model_history=classifier.fit(X_train,y_train,validation_split=0.33,batch_size=10,epochs=100,verbose=1)
```

```python
model_history.history.keys()
```

```python
y_pred=classifier.predict(X_test)
```

```python
# summarize history for accuracy
plt.plot(model_history.history['accuracy'])
plt.plot(model_history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
# summarize history for loss
plt.plot(model_history.history['loss'])
plt.plot(model_history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```

```python
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True,cmap="gist_earth")
```

```python
# Calculate the Accuracy
from sklearn.metrics import accuracy_score
score=accuracy_score(y_pred,y_test)
score
```
