### **Reading of CSV file**

```python
messages = pd.read_csv("/kaggle/input/sms-spam-collection-dataset/spam.csv", encoding='latin-1')
messages.dropna(how="any", inplace=True, axis=1)
messages.columns = ['label', 'message']
messages.head()
```

```python
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

# **Stemming**

```python
ps=PorterStemmer()
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
corpus
```

### **Bag of Words**

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500)
x=cv.fit_transform(corpus).toarray()
x
```

```python
y=pd.get_dummies(messages['label'])
y=y.iloc[:,1].values
y
```

### **Train test Split**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### **Model Selection**

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model=MultinomialNB().fit(X_train,y_train)
y_pred=spam_detect_model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

# **Lemmatization**

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmetizer=WordNetLemmatizer()
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['message'][i])
    review=review.lower()
    review=review.split()
    review=[lemmetizer.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
corpus
```

### **TF-IDF**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer(max_features=2000)
x=cv.fit_transform(corpus).toarray()
x[0]
```

```python
y=pd.get_dummies(messages['label'])
y=y.iloc[:,1].values
y
```

### **Train-Test Split**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

### **Model Selection**

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model=MultinomialNB().fit(X_train,y_train)
y_pred=spam_detect_model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```
