```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import sklearn

from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
```

```python
#----------Load Data Set
```

```python
dataset = pd.read_csv("../input/social-network-ads/Social_Network_Ads.csv")
```

```python
#-------------Separate Data Set into Numeric And Catagorical Columns for Analysis
```

```python
X=dataset.iloc[:,2:-1].values
y=dataset.iloc[:,-1].values
```

```python
#---------------Separate Training And Testing Data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.33,random_state=0)
```

```python
X_train
```

```python
y_train
```

```python
#---------------Train the Model

model = RandomForestClassifier()
```

```python
#----------------Fit the model for training data

model.fit(X_train, y_train)
```

```python
#-------------------Predict the Model on Test Data

y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
#---------------Find the Accuracy

Accuracy = accuracy_score(y_test, y_pred)
Accuracy
```

```python
#-----------------Find feture importances------------------------

model1 = RandomForestClassifier()
model1.fit(X_train, y_train)
```

```python
model1.feature_importances_
```

```python
feature_imp = pd.Series(data=model1.feature_importances_,index=['2','3' ]).sort_values(ascending=False)
feature_imp
```

```python
sns.barplot(x=feature_imp, y=feature_imp.index)
plt.ylabel('Feature Name')
plt.xlabel('Accuracy Score')
plt.title('Visulizing features')
plt.show()
```
