```python
city=["Bengaluru","Mumbai","Pune","Hyderabad"]
```

```python
plt.figure(figsize = (20, 30))
for i in enumerate(city):
    plt.subplot(2, 2, i[0]+1)
    sns.barplot(x=naukri_df[naukri_df['Location'].isin([i[1]])]['Job Title'].value_counts().nlargest(n=20).index,
                        y=naukri_df[naukri_df['Location'].isin([i[1]])]['Job Title'].value_counts().nlargest(n=20).values)
    plt.title(i[1])
    plt.xlabel("Location")
    plt.ylabel("Job Titles")
    plt.tight_layout(pad=3.0)
    plt.xticks(rotation = 90)
```

#### Insights from Graph:
- For Benguluru Location:
    The Top 20 jobs mostly consists of Software Engineer ,PHP Developer,Full Stack Developer, UI Developer,Sales Manager,
    HR Executive,Data Scientist,Business Analyst mostly related to IT Sector and Financial Fields.
    The Maximum vacancies of jobs is for Software Engineer.
- For Mumbai Location:
   The Top 20 jobs mostly consists of Business Development Manager,Business Development Executive,Graphic Designer,
   Java Developer,Sales Executive,Marketing Executive,Accounts Manager,Accountant,HR Recruiter 
   mostly related to Manageriol Roles and  IT Sector.
   The Maximum vacancies of jobs is for Business Development Manager . 
- For Pune Location: 
   The Top 20 jobs mostly consists of Java Developer,Business Development Executive,Senior Software Engineer,
   Technical Support Engineer,Technical Writer,Receptionist and Job Description which  used when there is no description given 
   by the recruiter. 
   The Maximum vacancies of jobs is for Java Developer. 
- For Hyderabad Location:
    The Top 20 jobs mostly consists of Job Description which  used when there is no description given 
    by the recruiter,Dot Net Developer,Android Developer,PHP Developer,IT Recruiter,Customer Support Executive,Sales Executive.
    The Maximum vacancies of jobs is for Job Description which means the recruiters in Hyderabad dont give much information
    about Job Description.
