```python
#Seperating feature and label variables
x = data.drop('quality',axis=1)
y = data['quality'].apply(lambda x:1 if x>=7 else 0)
#I have done binarisation of the target variable
#I'm treating quality as good(1) for all values above 6 & bad(0) for all values equal or below 6
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
model = RandomForestClassifier(random_state=1)
```

```python
model.fit(x_train,y_train)
```

```python
y_test_pred = model.predict(x_test)
y_train_pred = model.predict(x_train)
```

```python
accuracy_score(y_train_pred,y_train)
```

```python
accuracy_score(y_test_pred,y_test)
```
