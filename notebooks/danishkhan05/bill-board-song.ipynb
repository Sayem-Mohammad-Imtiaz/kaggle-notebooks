```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action='ignore')
%matplotlib inline
```

```python
data = pd.read_csv('/kaggle/input/billboard-the-hot-100-songs/charts.csv')
```

```python
#Converting the date column to datetime object
data.date = pd.to_datetime(data.date)
```

```python
#Creating year and month columns. It might be useful in EDA.
data['Year'] = pd.Series(val.year for val in data['date'])
```

```python
data['Month'] = pd.Series(val.month for val in data['date'])
```

```python
#last-week has some null values but we can't delete them. They indicate that these are the latest songs and
#hence were unranked in the previous week.
#Let's just impute 0 for all such cases.
data['last-week'] = np.where(data['last-week'].isnull(),0,data['last-week'])
```

```python
data.info()
```

```python
data['last-week'].value_counts()
```

```python
data.loc[data['last-week']==0,'song']
#All these songs were unrated last week
```

```python
#Let's see which artist's have the maximum number of songs
data['artist'].value_counts()[:50]
```

```python
#Shoeing the highest ranked song of each artist
data.groupby('artist').max()[['peak-rank','song']].sort_values(by='peak-rank')
```

```python
#Songs spending the highest times on the billboard
data.sort_values(by='weeks-on-board',ascending=False).iloc[:25,2:7]
```
