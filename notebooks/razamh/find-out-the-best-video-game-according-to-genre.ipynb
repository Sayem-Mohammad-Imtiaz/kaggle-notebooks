# BEST GAME ACCORING TO GENRE

```python
genres = vg['genre'].unique()
genres
```

```python
best_VG = {}

for genre in genres:
    #print(genre)
    genre_mask = vg['genre'] == genre
    gen = vg[genre_mask]
    #print(genre)
    VG = gen.sort_values("rank").iloc[0]['name']
    best_VG[genre] = VG
```

```python
best_VG
```

```python
Series = pd.Series(best_VG)
```

```python
pd.DataFrame(Series, columns = ['Games'])
```
