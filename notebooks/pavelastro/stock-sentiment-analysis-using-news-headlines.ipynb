```python
df=pd.read_csv('../input/stocknews/Combined_News_DJIA.csv',encoding="ISO-8859-1")
#utf-8 codec cant decode byte 0xa3 in position 16: invalid start byte
```

```python
df.head()
```

```python
df.tail()
```

```python
train=df[df['Date']<'20150101']
test=df[df['Date']>'20141231']
```

```python
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
#Renaming
list1=[i for i in range(25)]
new_Index=[str(i) for i in list1]
data.columns=new_Index
data.head(5)
```

```python
for index in new_Index:
    data[index]=data[index].str.lower()
data.head(1)
```

```python
' '.join(str(x) for x in data.iloc[1,0:25])
```

```python
headlines=[]
for row in range(0,len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row,0:25]))
```

```python
headlines[1
         ]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
#Converting all the sentences to Vectors
```

```python
# Using count Vectorizer to create Bag of Words
countvector=CountVectorizer(ngram_range=(2,2))
trainerdataset=countvector.fit_transform(headlines)
```

```python
trainerdataset[0] #sparse matrix
```

```python
#implementing randomforest classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(trainerdataset,train['Label'])
```

```python
test_transform=[]
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset=countvector.transform(test_transform)
predictions=randomclassifier.predict(test_dataset)
```

```python
test.loc[1611,:]
```

```python
#Checking the accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```

### Instead of count vectorization, using TFIDF to check how the accuracy is affected

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
tfidfvector=TfidfVectorizer(ngram_range=(2,2))
trainerdataset=tfidfvector.fit_transform(headlines) #Bag of words same steps
```

```python
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(trainerdataset,train['Label'])
```

```python
test_transform=[]
for row in range(0,len(test.index)):
    test_transform.append(' '.join(str(x) for x in test.iloc[row,2:27]))
test_dataset=tfidfvector.transform(test_transform)
predictions=randomclassifier.predict(test_dataset)
```

```python
predictions
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```
