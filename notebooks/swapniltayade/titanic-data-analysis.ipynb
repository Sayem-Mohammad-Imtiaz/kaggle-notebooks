```python
d = {'male':0 , 'female': 1}
data['Sex'] = data['Sex'].map(d)
data.head()
```

```python
o = {0:'male' , 1:'female'}
data = data.replace({'Sex':o})
data.head()
```

```python
as_fig = sns.FacetGrid(data,hue='Sex',aspect=5)
as_fig.map(sns.kdeplot,'Age',shade=True)
oldest = data['Age'].max()
as_fig.set(xlim=(0,oldest))
as_fig.add_legend()
```
