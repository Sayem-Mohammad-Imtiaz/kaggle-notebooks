```python
cualitativos= []

for i in df.columns:
    if df[i].dtypes== "object":
        cualitativos.append(i)
cualitativos
        
```

```python

cuantitativos= []
for i in df.columns:
    if i in cualitativos:
        None
    else:
        cuantitativos.append(i)
```

```python
#moda cualitativo
import scipy
from scipy import stats 

for i in cualitativos:              
    print(df[i].mode()[0])
```

```python
#rango cuantitativo
for i in cuantitativos:
    print("EL RANGO  ", i, "ES ", df[i].max() - df[i].min())
```

```python
# varianza cuantitativo

for i in cuantitativos:
    print("Varianza de", i, ":",np.var(np.array(df[i].values)))
```
