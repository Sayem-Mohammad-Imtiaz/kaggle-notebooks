```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('../input/iris-flower-dataset/IRIS.csv')
df.head()
```

```python
#first we try the normal encoder
encode = LabelEncoder()
```

```python
x = df.drop('species', axis = 1)
x.head()
```

```python
y = df.species
y.head()
```

```python
y_encoded = encode.fit_transform(y)
print(y_encoded)
```

```python
x_test, x_val, y_test, y_val = train_test_split(x,y_encoded,test_size=0.2,random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(x_test, y_test)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(classifier,filled=True)
plt.show()
```

```python
y_pred = classifier.predict(x_val)
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_val, y_pred)
cm
```

```python
from sklearn.tree import DecisionTreeClassifier
gini_c = DecisionTreeClassifier(criterion = 'gini', random_state = 0)
gini_c.fit(x_test, y_test)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(gini_c,filled=True)
plt.show()
```

```python
y_pred = gini_c.predict(x_val)
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_val, y_pred)
cm
```
