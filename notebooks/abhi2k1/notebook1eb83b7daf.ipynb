```python
data1=pd.read_csv("../input/hotel-booking-demand/hotel_bookings.csv")
data1.head()
```

```python
print("Shape of Data :",data1.shape)
print("Properties :",data1.info())
```

```python
data1.isna().sum()
```

```python
data1['children']=data1['children'].fillna(0)
data1['children'].isna().sum()
```

```python
temp=data1.iloc[-1,13]
data1['country']=data1['country'].fillna(temp)
data1['country'].isna().sum()
```

```python
temp1=data1['agent'].mean()
data1['agent']=data1['agent'].fillna(temp1)
data1['agent'].isna().sum()
```
