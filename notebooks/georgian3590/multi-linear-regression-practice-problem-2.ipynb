```python
import numpy as np 
import pandas as pd 
import matplotlib as mt
import seaborn as sns 

%matplotlib inline 
```

```python
data_input  = pd.read_csv("/kaggle/input/housesalesprediction/kc_house_data.csv")
```

```python
data_input.info()
```

```python
data_input.isnull().values.any()
```

```python
data = data_input.drop('date',axis = 1)
```

```python
data.head()
```

```python
data.count()
```

```python
sns.heatmap(data)
```

```python
data.corr()

```

```python
data_final = data.drop(['id','sqft_lot','condition','yr_built','yr_renovated','zipcode','long','sqft_lot15','floors'], axis=1)
```

```python
price = data_final.price
data = data_final.drop('price',axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import r2_score
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(data,price, test_size=0.33, random_state=42)
```

```python
lg = LinearRegression()
```

```python
lg.fit(X_train,Y_train)
```

```python
Y_predict = lg.predict(X_test)
```

```python
# first r2 with floors
r2_score(Y_test,Y_predict)
```

```python
# second r2 with floors
r2_score(Y_test,Y_predict)
```
