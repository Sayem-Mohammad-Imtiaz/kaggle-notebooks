```python
import numpy as np
import pandas as pd
import os
import statsmodels.api as sm
import scipy.stats as stats
import pylab

print(os.listdir("../input"))
```

```python
df = pd.read_csv('../input/NBA_player_of_the_week.csv')
```

```python
df = df['Weight']
```

```python
df.unique()
```

```python
df = df.apply(lambda s: float(s[:-2]) if 'kg' in s else int(s) * 0.453592) # unify kg and pounds
```

```python
np.sort(df.unique())
```

```python
df.describe()
```

```python
df.median()
```

```python
stats.variation(df)
```

```python
stats.kurtosis(df)
```

```python
stats.skew(df)
```

```python
df.plot.box()
```

```python
df.hist(bins=5)
```

```python
df.hist(bins=10)
```

```python
df.plot.density()
```

```python
probplot = sm.ProbPlot(df.values, stats.t, fit=True)
```

```python
_ = probplot.ppplot(line='45')
```

```python
_ = probplot.qqplot(line='45')
```
