```python
from lifelines import KaplanMeierFitter
kmf = KaplanMeierFitter()
```

```python
kmf.fit(df['Operation Year'], df['Survival'])
```

```python
df.Node.value_counts()
```

```python
kmf.plot_survival_function()
plt.show()
```

```python
from lifelines import WeibullFitter, ExponentialFitter, LogNormalFitter
wb = WeibullFitter().fit(df['Operation Year'], df['Survival'])
ex = ExponentialFitter().fit(df['Operation Year'], df['Survival'])
ln = LogNormalFitter().fit(df['Operation Year'], df['Survival'])
```

```python
models = [wb,ex,ln]
```

```python
print(wb.AIC_,ex.AIC_,ln.AIC_)
```

```python
from lifelines.plotting import qq_plot
import seaborn as sns
sns.set()
for model in [WeibullFitter(), ExponentialFitter(),LogNormalFitter()]:
    model.fit(df['Operation Year'], df['Survival'])
    qq_plot(model)
    plt.show()
```

```python
from lifelines import CoxPHFitter
cph = CoxPHFitter()
cph.fit(df = df, duration_col = 'Operation Year', event_col = 'Survival')
```
