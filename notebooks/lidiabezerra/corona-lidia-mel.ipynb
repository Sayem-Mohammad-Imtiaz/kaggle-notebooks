Esse vírus foi descoberto na província de Wuhan, na China, na data de 31 de dezembro, de 2019, por isso o víros ficou conhecido como "COVID-19".
    A transmissão de animal para humanos se deu por meio do contato com secreções e fluídos, ou seja, na forma inadequada de manuseio de cortes desses animais. Após a contaminação, os sintomas podem variar entre um resfriado comum e uma pneumonia severa, o que pode levar o indivíduo à dependência de um prespirador mecânico. Os principáis simtomas são:
* Tosse seca
* Dor de garganta
* falta de ar
    A transmissão ocorre entre uma pessoa saudável e uma doente, no entanto, se o ambiente possuir objetos contaminados, ou se no ar existir gotículas flutuantes a contaminação pode ocorrer, basta a pessoa tocar nos objetos e direcionar as mãos para os olhos, boca e nariz, ou respirar as gostículas existentes no arambiente. Portando, recomenda-se lavar as mãos com água e sabão (por, pelo menos, 20 segundos), tomar banho ao chegar em casa. Adicionalemente, recomenda-se o distanciamento social, devido à proximidade com a pessoa que possa espirrar ou falar cuspindo próximo a outra. **INICIANDO A SÉRIE TEMPORAL NO BRASIL**

```python
###INICIANDO AS ANÁLISES NO BRASIL
corona_brasil = corona[corona['country']=='Brazil']

corona_brasil.head()
```

```python
##Verificando número de casos 
corona_brasil.shape
##É possível observar que apenas 57 observações
```

```python
#Filtrar as colunas 
corona_brasil.drop(columns=[ 'death', 'recover', 'Last Update', 'country', 'cases_date'], inplace=True, axis=1)
corona_brasil.head()
```

```python
corona_brasil.index = corona_brasil['date']
corona_brasil.drop('date',inplace=True, axis=1)
corona_brasil.head()
```

```python
plt.figure(figsize=(40,20))
plt.plot(corona_brasil)
plt.xlabel('Data')
plt.ylabel('Confirmado')
plt.title('Data vs casos confirmados')
plt.xticks(rotation='vertical')
plt.show()
```

```python
mean = corona_brasil.rolling(window=12).mean()
m_std = corona_brasil.rolling(window=12).std()

ori = plt.plot(corona_brasil, color='blue', label='Original')
mm = plt.plot(mean, color='red',label='Média')
std = plt.plot(m_std, color='black', label='Desvio')
plt.legend(loc='best')
plt.title('Evolução dos casos de coronavírus')
  
plt.show()

result = adfuller(corona_brasil['confirm'])
print('ADF Statistic: {}'.format(result[0]))
print('p-value: {}'.format(result[1]))
print('Critical Values:')
for key, value in result[4].items():
   
    print('\t{}: {}'.format(key, value))
```

***APLICANDO O MODELO ARIMA***

```python
modelo = ARIMA(corona_brasil, order=(0, 1, 1)).fit()
print(modelo.summary())
```

```python
##Modelo preditivo para COVID-19
fig,ax = plt.subplots(figsize=(8,6))
fig = modelo.plot_predict(start='02-28-2020',end='04-18-2020',ax=ax)
plt.title('Previsão COVID_19')
plt.ylabel('Confirmados')
plt.xlabel('Data')
legend = ax.legend(loc='upper left')
```

****APLICANDO O MODELO NAIVE****

```python
#Modelo Naive sem sazonalidade
naive = ARIMA(corona_brasil, order=(0, 0, 0))
naive_treinado = naive.fit()
print(naive_treinado.summary())
```

```python
plt.rcParams.update({'figure.figsize':(9,3)})
naive_treinado.plot_predict(start=1,end=230)
plt.show()
```

```python
#Modelo naive com sazonalidade

snaive = ARIMA(corona_brasil, order=(0, 1, 0))
snaive_treinado = snaive.fit()
print(snaive_treinado.summary())
```

```python
#Treinando o modelo com sazonalidade
snaive_treinado.plot_predict(start=1,end=230)
plt.show()
```
