#### Explore correletion between customer ratings and other features

This would help us determine the factors or features that impact how customers rate the products.

```python
data.head()
```

```python
# let convert the object type columns

object_cols = list(data.dtypes[data.dtypes == object].index)

for i in object_cols:
    data[i] = data[i].astype('category')
```

```python
cat_columns = data.select_dtypes(['category']).columns
data[cat_columns] = data[cat_columns].apply(lambda x: x.cat.codes)
```

```python
data.head()
```

```python
data.columns
```

```python
features=['Warehouse_block', 'Mode_of_Shipment', 'Customer_care_calls',
          'Cost_of_the_Product', 'Prior_purchases','Product_importance', 
          'Gender','Reached.on.Time_Y.N']
```

```python
for f in features:
    related = data['Customer_rating'].corr(data[f])
    print("%s: %f" % (f,related))
```

```python
correlations = [data['Customer_rating'].corr(data[f]) for f in features ]
```

```python
features_df = pd.DataFrame({'attributes': features, 'correlation': correlations}) 
```

```python
features_df.plot(x ='attributes', y='correlation', kind = 'line',\
                    color='coral', figsize=(16,4), title='Customer Rating by Feature Correlation',\
                    linewidth=3.3, marker='.', markersize=20)                

plt.xlabel('Features')
plt.ylabel('Customer Ratings')
```

We see a high correlation between Customer Rating and Reach on Time, Prior Purchase, Customer Care Calls, Warehouse and cxost of Products. We can conclude that customers who are satisfied in these areas are likely to give the product better rating.
