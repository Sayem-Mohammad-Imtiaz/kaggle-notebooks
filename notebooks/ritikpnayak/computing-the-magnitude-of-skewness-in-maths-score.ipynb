# Computing the skewness

***What is Skewness?***

1. A property that describes the shape of a distribution
2. If the values extend farther to the right, it's "right skewed" and if the values extend left, it's "left skewed"

In our histograms, plotted above, both our histograms are left skewed. We can verify the same using a function that will compute the skewness of the data. ***In this case, the sign of the skewness must be negative because they are both left skewed.***

# 1. Using Moments

***Why use Moments?***

1. We can find the mean, median and variance/standard deviation using moments

```python
def RawMoment(xs, k):
    return sum(x ** k for x in xs) / len(xs)

def CentralMoment(xs, k):
    mean = RawMoment(xs, 1)
    return sum((x - mean) ** k for x in xs) / len(xs)

def StandardizedMoment(xs, k):
    var = CentralMoment(xs, 2)
    std = math.sqrt(var)
    return CentralMoment(xs, k) / std ** k

def Skewness(xs):
    return StandardizedMoment(xs, 3)
```

```python
print('Skewness in marks of female students: ', Skewness(female['math score'].values))
print('Skewness in marks of male students: ', Skewness(male['math score'].values))
```

# 2. Using Pearson's Median Coefficient

***Why use Pearson's Median Coefficient?***

1. More efficient
2. Based on the difference between the sample mean and the median

```python
def PearsonMedianCoeff(sample, xbar, median):
    gp = 3 * (xbar - median) / len(sample)
    return gp
```

```python
xbar_fem = female['math score'].mean()
xbar_male = male['math score'].mean()

median_fem = female['math score'].median()
median_male = male['math score'].median()
```

```python
print('Skewness in marks of female students: ', PearsonMedianCoeff(female['math score'].values, xbar_fem, median_fem))
print('Skewness in marks of male students: ', PearsonMedianCoeff(male['math score'].values, xbar_male, median_male))
```

***What does the magnitude tell us?***

1. The skewness in the marks of female students is more than that of the male students.
2. In both the cases the skewness is very small, indicating that both the distributions are nearly normal.

***What if the distributions were not normal?***

1. One explaination might be that the Maths exam was either too easy or too difficult.
2. Other explanation: The checking was either very strick or very loose.
