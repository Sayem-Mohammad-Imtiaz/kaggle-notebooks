# 2. Defining Covariance and Correlation

**Covariance** is the paired analogue of variance. The difference between variance and covariance is that, variance measures how a single variable deviates from its mean whereas **covariance measures how 2 variables vary alongside each other from their means**.

Positive Covaiance - When the corresponding values of x and y are either both above their means or both below their means.

Negative Covariance - When one is above its mean and the other below.

A large positive covariance means that x tends to be large when y is large and small when y is small.
A large negative covariance means that x tends to be small when y is large and vice versa.

Covariance, however, is hard to interpret, because:

1. It's units are the product of the inputs' units, which can be hard to make sense of (e.g. what is a "toilet for girls-ger of girls")

2. It's hard to interpret what counts as a "large" covariance.

I'll visualize the results using **scatter plot** which I think is the most convenient way to express correlation

# 2.1. Algorithm for covariance

```python
def de_mean(x):
    x_bar = mean(x)
    return [x_i - x_bar for x_i in x]

def covariance(x, y):
    n = len(x)
    return np.dot(de_mean(x), de_mean(y)) / (n - 1)
```

# 2.2. Algorithm for correlation

```python
def correlation(x, y):
    std_x = stdev(x)
    std_y = stdev(y)
    if std_x > 0 and std_y >0:
        return covariance(x, y) / std_x / std_y
    else:
        return 0
```
