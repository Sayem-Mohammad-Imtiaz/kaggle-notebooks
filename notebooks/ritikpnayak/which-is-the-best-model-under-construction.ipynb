```python
df = pd.read_csv('/kaggle/input/breast-cancer-wisconsin-data/data.csv')
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.iloc[:, 1: 12]
df.head()
```

```python
df.diagnosis.value_counts()
```

```python
df['diagnosis'] = df['diagnosis'].map(lambda x: 1 if x == 'M'
                                                 else 0)
```

```python
df.diagnosis.value_counts()
```

```python
X = df.copy()
y = X.pop('diagnosis')
```

```python
X.shape
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(X, y)
print(f'{lr.score(X, y):.4f}')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

dc = DecisionTreeClassifier()
dc.fit(X, y)
cv = cross_val_score(dc, X, y, cv = 10)
print(cv)
print()
print(sum(cv) / len(cv))
print()
print(cv.std())
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X, y)
cv = cross_val_score(knn, X, y, cv = 10)
print(cv)
print()
print(sum(cv) / len(cv))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = [{'min_samples_leaf' : [2, 4, 6, 8], 'max_features' : [2, 4, 6, 8]}]
grid_search = GridSearchCV(dc, param_grid, cv = 5)
grid_search.fit(X, y)
grid_search.best_params_
```

```python
dc = DecisionTreeClassifier(max_features = 8, min_samples_leaf = 8)
dc.fit(X, y)
cv = cross_val_score(dc, X, y, cv = 10)
print(cv)
print()
print(sum(cv) / len(cv))
print()
print(cv.std())
```
