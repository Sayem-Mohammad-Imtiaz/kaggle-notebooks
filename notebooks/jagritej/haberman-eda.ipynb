```python
# import lib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
haberman=pd.read_csv("../input/habermans-survival-data-set/haberman.csv", header=None, names=['AGE', 'OP_YEAR', 'AXILNODE','SURV_STATS'])
print(haberman.shape) #Number of the rows and columns
print(haberman.columns) #Understand the label in dataset
```

```python
print(haberman.head())
```

**DataSet Information**: Age of patient at time of operation (numerical),Patient's year of operation (year - 1900, numerical),Number of positive axillary nodes detected (numerical) Survival status (class attribute) 1 = the patient survived 5 years or longer 2 = the patient died within 5 years

```python
haberman["SURV_STATS"]=haberman["SURV_STATS"].map({1:"survived",2:"died"})
```

```python
plot=haberman.plot(kind='scatter' ,x='AGE', color='r',y='AXILNODE') #matplotlib 
plt.show()
```

```python
sns.set_style("whitegrid")   #scatterplot with seaborn
sns.FacetGrid(haberman,hue="SURV_STATS" ,height=4) \
.map(plt.scatter,"AGE" ,"OP_YEAR") \
.add_legend();
plt.show()
```

```python
sns.set_style("whitegrid")   #pairplot 
sns.pairplot(haberman ,hue="SURV_STATS", height=3)
plt.show()
```

**OBSERVATION 1**.In the plot points are overlapped. 2 .In OP_YEAR and AGE plot , patient's age from 0 to appxo 35 are survived.

```python
for x, label in enumerate(list(haberman.columns)[:-1]):
    sns.FacetGrid(haberman, hue="SURV_STATS", height=4)\
    .map(sns.distplot, label)\
    .add_legend()
    plt.show()
```

```python
plt.figure(figsize=(20,5))
for x, label in enumerate(list(haberman.columns)[:-1]):
    counts, bin_edges = np.histogram(haberman[label], bins=10, density=True)
    pdf = counts/sum(counts)
    plt.plot(bin_edges[1:], pdf) 
 
```

```python
for x, label in enumerate(list(haberman.columns)[:-1]):
    plt.subplot(1, 3, x+1)
    print("   "+label+"   ")
    counts, bin_edges = np.histogram(haberman[label], bins=10, density=True)
    print("Bin Edges: {}".format(bin_edges))
    pdf = counts/sum(counts)
    print("PDF: {}".format(pdf))
    cdf = np.cumsum(pdf)
    print("CDF: {}".format(cdf))
    plt.plot(bin_edges[1:],pdf,label='died:pdf')
    plt.plot(bin_edges[1:],cdf,label='died:cdf')
    plt.legend()
    plt.xlabel(label)
```

```python
haberman_died=haberman[haberman.SURV_STATS == 'died']
for x, label in enumerate(list(haberman.columns)[:-1]):
    counts, bin_edges = np.histogram(haberman_died[label], bins=10, density=True)
    pdf = counts/sum(counts)
    cdf = np.cumsum(pdf)
    plt.plot(bin_edges[1:],pdf,label='died:pdf')
    plt.plot(bin_edges[1:],cdf,label='died:cdf')
```

```python
haberman_survived=haberman[haberman.SURV_STATS == 'survived']
for x, label in enumerate(list(haberman.columns)[:-1]):
    counts, bin_edges = np.histogram(haberman_survived[label], bins=10, density=True)
    pdf = counts/sum(counts)
    cdf = np.cumsum(pdf)
    plt.plot(bin_edges[1:], pdf ,bin_edges[1:], cdf)
```

**Observation AGE** : Patients having age from 0 and 35 years, survived more also patients having age more than 75 years, survived less. OP_YEAR : scatterpoint are overlapped. AXILNODE : Patients having axilnode 0 to 2 survived more also having more than 2axilnode, survived less.


```python
fig,axes = plt.subplots(1, 3, figsize=(15, 5))
for y, label in enumerate(list(haberman.columns)[:-1]):
    sns.boxplot( x='SURV_STATS', y=label, data=haberman, ax=axes[y])
plt.show()
```

```python
fig,axes = plt.subplots(1, 3, figsize=(15, 5))
for y, label in enumerate(list(haberman.columns)[:-1]):
    sns.violinplot( x='SURV_STATS', y=label, data=haberman, ax=axes[y])
plt.show()
```

**OBSERVATION:***(FROM BOXPLOT AND VIOLIN PLOT )* AGE: Patients whose age was less than 52 years are observed to survived morethan five years. OP_YEAR :Patients whose op_year was less than 60 year are observed to survived less than patients whose op_year more than 65 op_year. AXILNODE: Only 25% percentiles of patients have axilnode from 0 to 2 and survived more and others patient having more axilnode.


**FINAL OBSERVATION** : 
**AGE** : There are 50% of patients whose survived is less than 5 year that have age < 52, Patients having age from 0 and 35 years, survived more also patients having age more than 75 years, survived less. 
**OP_YEAR :** There are 72% of patients whose survived is less than 5 year that have op_year < 65,Patients whose op_year was less than 60 year are observed to survived less than patients whose op_year more than 65 op_year. 
**AXILNODE :** There are 81% of patients whose survived is less than 5 year that have axilnode > 2 ,patients have axilnode from 0 to 2 and survived more and others patient having more axilnode
