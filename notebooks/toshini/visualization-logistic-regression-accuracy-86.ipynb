# Splitting Data into X and Y

```python
x = heart.drop(['target'], axis = 1)
y = heart['target']
```

# Testing and Training Data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size = 0.6, random_state = 2)
```

# Scaling our x_train and x_test 

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_1 = sc.fit_transform(x_train);x_train_1.shape
```

```python
x_test_1 = sc.fit_transform(x_test);x_test_1.shape
```

# Implementing Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression(random_state = 3)
```

```python
log.fit(x_train_1,y_train)
```

```python
y_pred = log.predict(x_test_1)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_test, y_pred)
acc
```

```python
#Training score
Train_score = log.score(x_train_1,y_train)
#Testing score
Test_score = log.score(x_test_1,y_test)
```
