Importing Libraries

```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
%matplotlib inline
```

```python
df=pd.read_csv("../input/us-accidents/US_Accidents_Dec20.csv")
```

```python
df.head()
```

Exploring Start Time Column

```python
df.Start_Time[:5]
```

```python
#Converting into date time format
df.Start_Time=pd.to_datetime(df.Start_Time)
```

```python
sns.distplot(df.Start_Time.dt.hour,bins=24, kde=False)
plt.grid()
```

This shows that most of the accidents happen during 7-9 followed by 15-17 hrs, possibly because it is the traveling time of most of the people, they go to office at that time and return from office

```python
sns.distplot(df.Start_Time.dt.dayofweek,bins=24,kde=False)
```

This shows less accidents happen on weekends

It might be possible that time of accidents are different on weekdays and weekends****

```python
sns.distplot(df.Start_Time.dt.hour[df.Start_Time.dt.dayofweek==6],bins=24,kde=False)
plt.grid()
```

```python
sns.distplot(df.Start_Time.dt.hour[df.Start_Time.dt.dayofweek==1],bins=24,kde=False)
```

There is different time of peek accidents on weekdays and weekends

# Summary
* Start Time and End Time show similar results
* There is different trend of accidens on weekdays and weekends
* On weekdays 7-9 is the peeks time of accidents maybe because people go to work and then in the evening when people come back from work 
* On weekends it follows a gaussian curve and most accidents are during 12-14 maybe because less people go to work and at that time people for for outing with family and friends
