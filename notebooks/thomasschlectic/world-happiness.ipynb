# **Does Money = Happiness?**

```python
print(Happiness_Overall.corr())
plt.title("Variable Correlations", fontsize = 25)
sns.heatmap(Happiness_Overall.corr(), cmap="coolwarm", annot = True)
```

It looks like there is a strong coorelation between GDP and happiness lets take a look at this 

```python
from sklearn import datasets, linear_model
```

```python
#building a simple linear model 
wealthy_model = linear_model.LinearRegression()
wealthy_model.fit(Happiness_Overall[['Score']],Happiness_Overall[['GDP']])
print ("Intercept value " , wealthy_model.intercept_)
print ("coefficient" , wealthy_model.coef_)

plt.scatter(Happiness_Overall[['Score']],Happiness_Overall[['GDP']],color='blue')
plt.plot(Happiness_Overall[['Score']],wealthy_model.predict(Happiness_Overall[['Score']]),color='red',linewidth=4)
#plt.xticks(())
#plt.yticks(())

print("Formula for Happiness = " ,  wealthy_model.coef_ , "GDP +", wealthy_model.intercept_)
plt.show()
```

```python
#building a simple linear model 
corruption_model = linear_model.LinearRegression()
corruption_model.fit(Happiness_Overall[['Score']],Happiness_Overall[['Freedom']])
print ("Intercept value " , wealthy_model.intercept_)
print ("coefficient" , wealthy_model.coef_)

plt.scatter(Happiness_Overall[['Score']],Happiness_Overall[['Freedom']],color='red')
plt.plot(Happiness_Overall[['Score']],corruption_model.predict(Happiness_Overall[['Score']]),color='blue',linewidth=4)
#plt.xticks(())
#plt.yticks(())

print("Formula for Happiness = " ,  wealthy_model.coef_ , "GDP +", wealthy_model.intercept_)
plt.show()
```
