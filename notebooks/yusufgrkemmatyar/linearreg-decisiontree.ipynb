```python
df = pd.read_csv('../input/restaurant-business-rankings-2020/Top250.csv')
print(plt.style.available) # look at available plot styles
plt.style.use('ggplot')
```

```python
x = df.YOY_Sales.values.reshape(-1,1)
y = df.Units.values.reshape(-1,1)

plt.scatter(x,y)
plt.ylabel("Units")
plt.xlabel("Sales")

lr = LinearRegression()

lr.fit(x,y)


y_head = lr.predict(x)

plt.plot(x,y_head,color = "blue", label = "black")  
plt.show()
```

```python


lr.predict(np.array(15000).reshape(-1,1))
print('R^2 score: ',lr.score(x, y))

tree_reg = DecisionTreeRegressor()
tree_reg.fit(x,y)

x_ = np.arange(min(x),max(x),0.01).reshape(-1,1)
y_head = tree_reg.predict(x_)

plt.scatter(x,y,color="black")
plt.plot(x_,y_head,color="red")
plt.xlabel("Sales")
plt.ylabel("Units")
plt.show()
```
