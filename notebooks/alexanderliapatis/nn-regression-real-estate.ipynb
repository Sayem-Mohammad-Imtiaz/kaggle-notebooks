```python
import pandas as pd
df=pd.read_csv('/kaggle/input/real-estate-price-prediction/Real estate.csv')
```

```python
df.head()
```

```python
df=df.drop('X1 transaction date',axis=1)
df.head()
```

```python
len(df)
```

```python
df=df.drop('No',axis=1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop('Y house price of unit area',axis=1),
                                                            df['Y house price of unit area'],
                                                               test_size=0.2,random_state=30)
```

```python
X_train 
```

```python
y_train
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense,Dropout,BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping 
```

```python
model=Sequential()

model.add(Dense(9,activation='relu'))
model.add(BatchNormalization())

model.add(Dense(6,activation='relu'))
model.add(BatchNormalization())

model.add(Dense(3,activation='relu'))
model.add(BatchNormalization())

model.add(Dense(1,activation='linear'))

model.compile(loss='mae',optimizer='adam')
```

```python
early_stop=EarlyStopping(monitor='val_loss' , mode='min' , verbose=1 , patience=80 , restore_best_weights=True)
model.fit(x=X_train,y=y_train,epochs=500,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
import matplotlib.pyplot as plt
model_loss=pd.DataFrame(model.history.history)
model_loss.plot()
plt.xlabel('Epochs')
```

```python
model.evaluate(X_test,y_test,verbose=0)
```

```python
model.evaluate(X_train,y_train,verbose=0)
```

```python
len(X_test)
```

```python
prediction=model.predict(X_test)
predictions=pd.Series(prediction.reshape(len(X_test),))
```

```python
import numpy as np
y_test=np.array(y_test)
y_test=pd.Series(y_test)
y_test
pred_df=pd.DataFrame(y_test,columns=['True Y'])
```

```python
pred_df=pd.concat([pred_df,predictions],axis=1)
pred_df.columns=(['True Y','Prediction'])
import seaborn as sns
sns.scatterplot(data=pred_df,x='True Y',y='Prediction')
```

```python
sns.regplot(data=pred_df,x='True Y',y='Prediction')
```
