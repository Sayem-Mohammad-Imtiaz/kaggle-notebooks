```python
from pycaret.internal.pycaret_experiment import TimeSeriesExperiment
from sktime.utils.plotting import plot_series
```

```python
GSPTSE.index = pd.DatetimeIndex(GSPTSE["Date"])
GSPTSE.drop(["Date"],axis=1,inplace=True)
GSPTSE.head()
```

```python
GSPTSE = GSPTSE.asfreq('D')
GSPTSE.index
```

```python
GSPTSE=GSPTSE['1996-04-01':]
GSPTSE.head()
```

```python
GSPTSE=GSPTSE.fillna(method='ffill')
GSPTSE.head()
```

```python
data=GSPTSE.Close
data[:5]
```

```python
import matplotlib.pyplot as plt
_ =plot_series(data)
plt.xticks(rotation=90)
plt.grid()
```

```python
exp = TimeSeriesExperiment()
exp.setup(data=data,session_id=42,fh=365)
```

```python
exp.compare_models()
```
