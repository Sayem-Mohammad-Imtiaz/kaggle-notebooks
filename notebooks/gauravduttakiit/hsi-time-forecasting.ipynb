```python
from pycaret.internal.pycaret_experiment import TimeSeriesExperiment
from sktime.utils.plotting import plot_series
```

```python
HSI.index = pd.DatetimeIndex(HSI["Date"])
HSI.drop(["Date"],axis=1,inplace=True)
HSI.head()
```

```python
HSI = HSI.asfreq('D')
HSI.index
```

```python
HSI=HSI['1996-04-01':]
HSI.head()
```

```python
HSI=HSI.fillna(method='ffill')
HSI.head()
```

```python
data=HSI.Close
data[:5]
```

```python
import matplotlib.pyplot as plt
_ =plot_series(data)
plt.xticks(rotation=90)
plt.grid()
```

```python
exp = TimeSeriesExperiment()
exp.setup(data=data,session_id=42,fh=365)
```

```python
exp.compare_models()
```
