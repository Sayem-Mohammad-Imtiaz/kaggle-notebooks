```python
from pycaret.internal.pycaret_experiment import TimeSeriesExperiment
from sktime.utils.plotting import plot_series
```

```python
N225.index = pd.DatetimeIndex(N225["Date"])
N225.drop(["Date"],axis=1,inplace=True)
N225.head()
```

```python
N225 = N225.asfreq('D')
N225.index
```

```python
N225=N225['1996-04-01':]
N225.head()
```

```python
N225=N225.fillna(method='ffill')
N225.head()
```

```python
data=N225.Close
data[:5]
```

```python
import matplotlib.pyplot as plt
_ =plot_series(data)
plt.xticks(rotation=90)
plt.grid()
```

```python
exp = TimeSeriesExperiment()
exp.setup(data=data,session_id=42,fh=365)
```

```python
exp.compare_models()
```
