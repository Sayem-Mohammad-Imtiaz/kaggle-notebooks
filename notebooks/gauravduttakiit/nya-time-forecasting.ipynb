```python
from pycaret.internal.pycaret_experiment import TimeSeriesExperiment
from sktime.utils.plotting import plot_series
```

```python
NYA.index = pd.DatetimeIndex(NYA["Date"])
NYA.drop(["Date"],axis=1,inplace=True)
NYA.head()
```

```python
NYA = NYA.asfreq('D')
NYA.index
```

```python
NYA=NYA['2001-04-02':]
NYA.head()
```

```python
NYA=NYA.fillna(method='ffill')
NYA.head()
```

```python
data=NYA.Close
data[:5]
```

```python
import matplotlib.pyplot as plt
_ =plot_series(data)
plt.xticks(rotation=90)
plt.grid()
```

```python
exp = TimeSeriesExperiment()
exp.setup(data=data,session_id=42,fh=365)
```

```python
exp.compare_models()
```
