```python
dataset=pd.read_csv('../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv')
dataset
```

```python
dataset.isnull().sum()
```

```python
y=dataset['DEATH_EVENT']
X=dataset.drop('DEATH_EVENT',axis=1)
X
```

# Splitting the data into train and test data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)

```

# Hyperparameter Tuning Using RandomizedSearchCV

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

# For Decision Tree Classifer

```python
from sklearn.model_selection import RandomizedSearchCV
rs=RandomizedSearchCV(DecisionTreeClassifier(),{
            'criterion':['gini','entropy'],
            'min_samples_leaf':[x for x in range(1,6)],
            'min_samples_split':[x for x in range(2,6)]
        },cv=3,return_train_score=False)
rs.fit(X_train,y_train)
rs.cv_results_
df=pd.DataFrame(rs.cv_results_)
df
```

```python
rs.best_score_
```

```python
model1=DecisionTreeClassifier(criterion='entropy',min_samples_split=4,min_samples_leaf=1)
model1.fit(X_train,y_train)
```

```python
model1.score(X_test,y_test)
```

# For Random Forest Classifier

```python
rs1=RandomizedSearchCV(RandomForestClassifier(),{
            'n_estimators':[x for x in range(10,100,10)],
            'max_depth':[20,25,30,35,40,45,50],
            'min_samples_leaf':[2,3,4,5],
            'min_samples_split':[x for x in range(2,11)]
        },cv=3,return_train_score=False)
rs1.fit(X_train,y_train)
rs1.cv_results_
df1=pd.DataFrame(rs1.cv_results_)
df1
```

```python
rs1.best_score_
```

```python
model2=RandomForestClassifier(n_estimators=30,max_depth=50,min_samples_leaf=4,min_samples_split=8)
model2.fit(X_train,y_train)
```

**SCORE PREDICTION**

```python
model2.score(X_test,y_test)
```
