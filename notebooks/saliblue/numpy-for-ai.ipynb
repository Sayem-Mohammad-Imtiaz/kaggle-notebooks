```python
z = np.arange(1, 20, step=3) #np.arange(): --> np.arange(start,stop,step)
print(z)
```

*np.linespace* :note that the maximum value is **included**, contrary to **arange**

```python
h = np.linspace(1, 2, num=5)# np.linespace: -->step = stop-start/num 
print(h)
```

```python
g = np.logspace(1, 5, num=5) # optinal base = 10 ,this examp: 10^1 --10^5 (start = 10, stop=100000 , num = 5) 
print(g)
```

```python
print(np.logspace(1, 5, num=5, base=2)) #this exmple: base =2 (start = 2,stop = 32, num = 5)
```
