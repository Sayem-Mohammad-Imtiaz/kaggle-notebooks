# **Analysis based on Age**

```python
#Youngest passenger who survived
test_titanic[test_titanic['Survived']==1]['Age'].min()
```

```python
#Oldest passenger who survived
test_titanic[test_titanic['Survived']==1]['Age'].max()
```

```python
#Combining Sibsp and Parch to form the family column
test_titanic['Family'] = test_titanic['SibSp'] + test_titanic['Parch']
test_titanic.head()
```

```python
#Grouping ages for all passengers in titanic
groups = ['0-9', '10-19', '20-29', '30-39', '40-49', '50-59', '60-69', '70-79']
test_titanic['Age_group'] = pd.cut(test_titanic.Age, range(0, 81, 10), right = False, labels = groups)
test_titanic.head()
```

```python
#Dropping sibsp, parch columns
test_titanic.drop(['SibSp'], axis = 1, inplace = True)
test_titanic.drop(['Parch'], axis = 1, inplace = True)
test_titanic.head()
```

```python
#Visualising passengers per age group per gender 
sns.countplot(x = 'Age_group', data = test_titanic, hue = 'Sex')
```

# **Mapping data**

```python
#Mapping survived vs Non survived passengers
survived_map = {0: 'No', 1: 'Yes'}
test_titanic['Survived'] = test_titanic['Survived'].map(survived_map)

# PClass map
pclass_map = {1: 'Upper Class', 2: 'Middle Class', 3: 'Lower Class'}
test_titanic['Pclass'] = test_titanic['Pclass'].map(pclass_map)

# Embarkation port map
port_map = {'S': 'Southampton', 'C': 'Cherbourg','Q':'Queenstown'}
test_titanic['Embarked'] = test_titanic['Embarked'].map(port_map)

test_titanic
```

```python
#Fare with respect to age group and gender
sns.barplot(x = 'Age_group', y = 'Fare', data = test_titanic, hue = 'Sex')
```

```python
#Fare with respect to class
sns.barplot(x = 'Age_group', y = 'Fare', data = test_titanic, hue = 'Pclass')
```

```python
#Removing all dead people
indexNames = test_titanic[test_titanic['Survived'] == 'No' ].index
# Delete these row indexes from dataFrame
test_titanic.drop(indexNames,inplace=True)
test_titanic.head()
```

```python
#Age group of survivors
sns.countplot(x = 'Age_group', data = test_titanic)
```

```python
#Fare of the survivors with respect to age group
sns.barplot(x = 'Age_group', y = 'Fare', data = test_titanic, palette = 'magma')
```

```python
#Boxplot for the above
sns.boxplot(x = "Age_group", y = "Fare", data = test_titanic, palette = 'rainbow')
```

```python
#Plot for survivors from the places they embarked
sns.countplot(x = 'Embarked', data = test_titanic)
```
