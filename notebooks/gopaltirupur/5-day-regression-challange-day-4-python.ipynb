Let us Try to predict the Income from the other Variables
From the above heatmap it is clear that,

1. Income is positively coorelated with 
" SelfEmployed	Employed	WorkAtHome	CensusTract	White	Citizen	IncomeErr	IncomePerCap	IncomePerCapErr	Professional	Drive	Walk	OtherTransp "

2. Income is negatively coorelated with
" Unemployment	MeanCommute	Hispanic		Black	Poverty	ChildPoverty	Service	Office	Construction	Production	Transit"

Income is directly dependent on the IncomePerCap, IncomePerCapErr, IncomeErr. Since we need to predict the Income we are assuming that the above variables also unknown along with the Income variable

" SelfEmployed	Employed	WorkAtHome	CensusTract	White	Citizen		Professional	Drive	Walk	OtherTransp  Unemployment	MeanCommute	Hispanic		Black	Poverty	ChildPoverty	Service	Office	Construction	Production	Transit"

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

def drawModel(reg,x,y):  
    reg.fit(x,y)
    y_pred = reg.predict(x)
    residual = y - y_pred
    plt.scatter(y_pred,residual)       
```

```python
# Now Let's Try to predict the Income using the selected Variables
# Removing ' CensusTract '
x = nyc_census_new[["SelfEmployed","Employed","WorkAtHome","White","Citizen","Professional","Drive","Walk","OtherTransp","Unemployment","MeanCommute","Hispanic","Black","Poverty","ChildPoverty","Service","Office","Construction","Production","Transit"]]
y=nyc_census_new['Income']
display(x.head())
drawModel(LinearRegression(),x,y)
```

```python
#PAIR PLOT FOR UNDERSTANDING THE POSSIBLE RELATIONSHIP BETWEEN THE SELECTED VARIABLES
# SELECTING ONLY THE HIGHLY CORELATED VARIABLES FOR ANALYSIS
nyc_census_new = nyc_census.fillna(nyc_census.mean())
corr = nyc_census_new[['Poverty','Income','IncomePerCap','Professional','SelfEmployed','White']].corr()
corr.head()
f,ax = plt.subplots(figsize=(10,10))
sns.heatmap(corr,annot=True,fmt='g',cmap='viridis',square=True,ax=ax,xticklabels=1, yticklabels=1)
```

```python
x = nyc_census_new[['Professional','Poverty','IncomePerCap']]
y=nyc_census_new['Income']
```

```python
from sklearn.linear_model import LinearRegression
drawModel(LinearRegression(),x,y)
```

```python
from sklearn.linear_model import ARDRegression
drawModel(ARDRegression(),x,y)
```

```python
from sklearn.linear_model import HuberRegressor
drawModel(HuberRegressor(),x,y)
```
