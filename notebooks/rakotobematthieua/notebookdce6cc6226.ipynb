```python
import pandas as pd

sf_salaries = pd.read_csv("../input/sf-salaries/Salaries.csv")
```

```python
#Grouper les Jobs en trouvant la somme des salaires en comptants juste basePay et OtherPay
new_salaries = sf_salaries.loc[:, ['JobTitle', 'BasePay', 'OtherPay']]
new_salaries.assign(somme_salaire_wtt = lambda x: x['BasePay'] + x['OtherPay']).groupby(['JobTitle']).sum().loc[:, ['somme_salaire_wtt']]

```

```python
#Grouper la somme de TotalPayBenefits de chaque jobs
new_salaries2 = sf_salaries.loc[:, ['JobTitle', 'TotalPayBenefits']].groupby(['JobTitle']).TotalPayBenefits.sum()
new_salaries2
```

```python
#Grouper len min et max de TotalPayBenefits par Job
new_salaries3 = sf_salaries.loc[:, ['JobTitle', 'TotalPayBenefits']].groupby(['JobTitle']).TotalPayBenefits.agg([len, min, max])
new_salaries3
```
