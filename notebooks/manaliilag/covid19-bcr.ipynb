```python
import pandas as pd  #data reading and processing 
import os             #reading from specific address
import matplotlib.pyplot as plt
import numpy as np
from datetime import *
import bar_chart_race as bcr
```

```python
data_corona = pd.read_csv("../input/covid19-in-india/covid_19_india.csv")
data_corona.head()
```

```python
data_corona.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational","Cured","Deaths"],axis=1,inplace=True)
```

```python
data_corona.head()
```

```python
Datess = []
for i in list(data_corona.Date):
    d,m,y = i.split('/')
    Datess.append(date(int(y)+2000,int(m),int(d)))
data_corona["Date"] = Datess
```

```python
data_corona.head(20)
```

```python
data_corona.drop(data_corona[data_corona['Date'] < date(2020,3,1)].index, inplace = True)
```

```python
data_corona.head()
```

```python
covidindia = data_corona.pivot_table(index="Date",columns="State/UnionTerritory",values="Confirmed",aggfunc="sum")
```

```python
covidindia.head()
```

```python
covidindia = covidindia.fillna(0)
```

```python
covidindia.head()
```

```python
bcr.bar_chart_race(df=covidindia,title="Covid 19 confirmed cases",n_bars=10,filter_column_colors=True,figsize=(5,3))
```
