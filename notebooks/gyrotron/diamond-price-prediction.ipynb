```python
df_raw = pd.read_csv("../input/diamonds/diamonds.csv", index_col=0)
df_raw.head()
```

```python
pp.ProfileReport(df_raw)
```

```python
#not in custom order
#df_raw["cut"].unique()
#df_raw["cut"].astype("category").cat.codes
cut_class_dict = {"Fair": 1, "Good": 2, "Very Good": 3, "Premium": 4, "Ideal": 5}
clarity_dict = {"I3": 1, "I2": 2, "I1": 3, "SI2": 4, "SI1": 5, "VS2": 6, "VS1": 7, "VVS2": 8, "VVS1": 9, "IF": 10, "FL": 11}
color_dict = {"J": 1, "I": 2, "H": 3, "G": 4, "F": 5, "E": 6, "D": 7}

df_raw["cut"] = df_raw["cut"].map(cut_class_dict)
df_raw["clarity"] = df_raw["clarity"].map(clarity_dict)
df_raw["color"] = df_raw["color"].map(color_dict)

df_raw.head()
```

```python
%matplotlib notebook
import seaborn as sns
import matplotlib.pyplot as plt

corr = df_raw.corr()

# plot the heatmap
sns.heatmap(corr, 
        xticklabels=corr.columns,
        yticklabels=corr.columns)
```

```python
import sklearn
from sklearn import svm, preprocessing

df = sklearn.utils.shuffle(df_raw)

X = df.drop("price", axis=1).values
X = preprocessing.scale(X)
y = df["price"].values

test_size = 200

X_train = X[:-test_size] #upto last 200
y_train = y[:-test_size]


X_test = X[-test_size:]
y_test = y[-test_size:]

clf = svm.SVR(kernel="linear")
clf.fit(X_train, y_train)
```

## Linear SVR

```python
clf.score(X_test, y_test)
```

```python
for X,y in zip(X_test, y_test):
    print(f"Model: {clf.predict([X])[0]}, Actual: {y}")
```

```python
clf = svm.SVR(kernel="rbf")
clf.fit(X_train, y_train)
```

## RBF SVR

```python
clf.score(X_test, y_test)
```

```python
for X,y in zip(X_test, y_test):
    print(f"Model: {clf.predict([X])[0]}, Actual: {y}")
```
