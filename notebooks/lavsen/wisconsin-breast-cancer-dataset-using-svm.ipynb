```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn import preprocessing
%matplotlib inline
import matplotlib.pyplot as plt

```

```python
df=pd.read_csv("../input/data.csv",header = 0)
```

```python
df.head()
```

```python
df.drop('id',axis=1,inplace=True)
df.drop('Unnamed: 32',axis=1,inplace=True)
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1,'B':0})
```

```python
df.head()
```

```python
data_df= list(df.columns[1:31]) 
data_df_main = df.loc[:,data_df]
```

```python
data_df_main
```

```python
X = data_df_main
y = df['diagnosis']
```

```python
svm_model = SVC()

parameters = [
              {'C': [1, 10, 100, 1000], 
               'kernel': ['linear']
              },
              
 ]
```

```python
grid_svm = GridSearchCV(svm_model, parameters, cv=20, scoring="accuracy")
grid_svm.fit(X,y)
```

```python
print(grid_svm.best_score_)
```

**Using SVM with all the features , there is accuracy of 0.95 when using linear kernel. After working on feature selection and different kernel selection , the accuracy might be improved further.**
