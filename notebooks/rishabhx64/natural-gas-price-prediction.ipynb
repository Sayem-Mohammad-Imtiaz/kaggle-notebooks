# Reading dataset

```python
dataset = pd.read_csv('/kaggle/input/natural-gas-price-forecasting/daily_csv.csv')
dataset1 = pd.read_csv('/kaggle/input/wti-data/wti-daily_csv(1).csv')
```

```python
dataset.head()
dataset1.head()
```

```python
dataset.info()
dataset1.info()
```

```python
dataset['year'] = pd.DatetimeIndex(dataset['Date']).year
dataset['month'] = pd.DatetimeIndex(dataset['Date']).month
dataset['day'] = pd.DatetimeIndex(dataset['Date']).day
dataset1['year'] = pd.DatetimeIndex(dataset1['Date']).year
dataset1['month'] = pd.DatetimeIndex(dataset1['Date']).month
dataset1['day'] = pd.DatetimeIndex(dataset1['Date']).day
```

```python
dataset
dataset1
```

# Droping Columns

```python
dataset.drop('Date', axis=1, inplace=True)
dataset1.drop('Date', axis=1, inplace=True)
```

# Checking for null values

```python
dataset.isnull().any()
dataset1.isnull().any()
```

```python
dataset.isnull().sum()
dataset1.isnull().sum()
```

# **Handling Missing values**

```python
dataset['Price'].fillna(dataset['Price'].mean(),inplace=True)
dataset1['Price'].fillna(dataset1['Price'].mean(),inplace=True)
```

Now, Split the dataset into X(independent variable) and Y(dependent variable)

```python
x=dataset.iloc[:,1:4].values #inputs
y=dataset.iloc[:,0:1].values #output price only
```

```python
x
```

```python
y
```

Split data into train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
#import decision tree regressor
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
#fitting the model or training the model
dtr.fit(x_train,y_train)
```

```python
y_pred=dtr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
accuracy=r2_score(y_test,y_pred)
accuracy
```
