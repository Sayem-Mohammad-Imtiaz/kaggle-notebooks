```python
df_2019 = pd.read_csv('/kaggle/input/world-happiness/2019.csv',index_col='Country or region')
# changing the name of the common columns for better understanding
df_2019.rename(columns={'Overall rank':'Happiness Rank','Country or region':'Country' ,'Score':'Happiness Score' , 'GDP per capita':'Economy (GDP per Capita)'
                        ,'Healthy life expectancy':'Health (Life Expectancy)','Freedom to make life choices':'Freedom' ,'Perceptions of corruption':
                        'Trust (Government Corruption)',},inplace=True)
# columns= Country or region,Happiness Rank , Happiness Score , Economy (GDP per Capita),Health (Life Expectancy),Freedom ,
#                   Trust (Government Corruption) ,Generosity
```

It would be better to divide the all the countries into the quartile. It will help us to visualise the data and will also help us to get some quick estimate by just looking the figure.

```python
# let's divide the ranks into quartile
quartile=['Forth','Third','Second','First']
df_2019['Quartile'] = pd.qcut(df_2019['Happiness Score'],q=4,labels=quartile)
df_2019.head()
```

Let's get the summery of the data set and view how it is distributed

```python
df_2019.describe()
```

From the above summery, we can see that the mean and median are almost equal for all the columns except Trust(Government Corruption). In the trust column their is a big difference.  

```python
# let's see how the columns are related to the each other
sns.pairplot(df_2019 , hue ='Quartile')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
corr_max = df_2019.corr()
corr_max
plt.figure(figsize=(12,10))
sns.heatmap(corr_max,annot = True,cmap=plt.cm.autumn,linewidth=1,linecolor='White')
plt.show()
```

It seems that *Generosity* does not have much impact on the data set, which is very interesting because it is generally consider a important part of happy nation.
*Freedom* is also weakly related to *Health*. It is also very intriguing that *Social Support* and *Happiness Rank* is not highly correlated with *Trust (Government Corruption)* which can be interpret that their are other sources that contribute towart the *Social Support* and countrymens can also be somewhat happy with *Trust Issues* :p with the government.  

```python
sns.scatterplot(x=df_2019['Economy (GDP per Capita)'],y=df_2019['Freedom'] , hue=df_2019['Quartile'] )
plt.title('Economy (GDP per Capita) vs Freedom' )
plt.show()
```

Freedom is not always a good indicator of the economic condititon of a country.It has been estimate from the above scatter plot that their are many country that is doing good in term of *Economy* but not in the *Freedom* and those are in the top 25% of the *Happiness Rank* 

I lives in India. So I am courious to know that how my country is doing in comperison of the top 10 country.

```python
df_top10_to_India = df_2019.head(10).append(df_2019.loc[(df_2019.index=='India')])
df_top10_to_India.drop(columns=['Happiness Rank','Quartile'], inplace=True)
```

```python
for a in df_top10_to_India.columns:
    plt.figure(figsize=(15,5))
    sns.barplot(x=df_top10_to_India.index,y=df_top10_to_India[a].values)
    plt.title(a)
    plt.show()
```
