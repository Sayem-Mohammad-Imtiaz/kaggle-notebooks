
World Happiness Report is yearly publication of UN. It contains articles and rankings of national happiness, based on respondent ratings of their own lives. In this notebook, I will drive meaning and comprehensive insight of that and also find how they are correlated to eaching other with the help the various visualization liabrary.

```python
#importing liabrary and data
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots


df=pd.read_csv('../input/world-happiness-report-2021/world-happiness-report-2021.csv')
previous_df=pd.read_csv('../input/world-happiness-report-2021/world-happiness-report.csv')
```

```python
df.head()
```

I will mainly focus on the below parameter.

**Ladder score** - Total score achived by the countries based of the parameter. 10 is given to the best life and 0 is for worst life.

**Logged GDP per capita** - GDP per capita is given in the logarithm form so that it can be compair with the other parameter.

**Social support** - This parameter is usually self explanatory. How people see the support provided by the socity including government and other institution. it is measured from 0 to 1.

**Healthy life expectancy** - How long they are expected to live. It is given in the years.

**Freedom to make life choices** - This parameter is also self explanatory. It is measured between 0 to 1.

**Generosity** - It measured how helping respondant is. It range from -1 to +1  

**Perceptions of corruption** -  This one explains how people of the country see honesty of their government.It range from 0 to 1

It would be better to divide the all the countries into 4 group based on the Laddder Score.

**First** - Top 25% countries.

**Second** - From Top 25% to 50% countries.

**Third** - From Top 50% to 75% countries.

**Forth** - From Top 75% to 100% countries.

I will name this column Quartile.
It will help us to visualise the data and will also help us to get some quick estimate by just looking the figure.

```python
quartile=['Forth','Third','Second','First']
df['Quartile'] = pd.qcut(df['Ladder score'],q=4,labels=quartile)
df.drop(columns=['Standard error of ladder score', 'upperwhisker', 'lowerwhisker','Explained by: Log GDP per capita', 'Explained by: Social support',
         'Explained by: Healthy life expectancy','Explained by: Freedom to make life choices',
         'Explained by: Generosity', 'Explained by: Perceptions of corruption','Dystopia + residual'],inplace=True)
```


# Top and Bottom 5 countries

```python
top_bottom_5=df.head().append(df.tail())
plt.figure(figsize=(10,8))
plt.style.use('fivethirtyeight')
sns.barplot(data=top_bottom_5,y='Country name',x='Ladder score')
plt.tight_layout()
plt.show()
```

# Analysis On the Regional Level

```python
plt.figure(figsize=(10,8))
plt.style.use('fivethirtyeight')
sns.boxplot(x=df['Regional indicator'],y=df['Ladder score'])
plt.tight_layout()
plt.xticks(rotation=75)
plt.show()
```

```python
reg_df = df.groupby(['Regional indicator'])

plt.style.use('fivethirtyeight')
fig,ax=plt.subplots(nrows=7,ncols=1,figsize=(20,30),sharex=True)
sns.barplot(y=reg_df['Ladder score'].mean().values,x=reg_df['Ladder score'].mean().index,ax=ax[0])
ax[0].axhline(df['Ladder score'].mean(),linestyle='--',color='black')
ax[0].set_title('Ladder score',fontsize=30)
sns.barplot(y=reg_df['Logged GDP per capita'].mean().values,x=reg_df['Logged GDP per capita'].mean().index,ax=ax[1])
ax[1].axhline(df['Logged GDP per capita'].mean(),linestyle='--',color='black')
ax[1].set_title('Logged GDP per capita',fontsize=30)
sns.barplot(y=reg_df['Social support'].mean().values,x=reg_df['Social support'].mean().index,ax=ax[2])
ax[2].axhline(df['Social support'].mean(),linestyle='--',color='black')
ax[2].set_title('Social support',fontsize=30)
sns.barplot(y=reg_df['Healthy life expectancy'].mean().values,x=reg_df['Healthy life expectancy'].mean().index,ax=ax[3])
ax[3].axhline(df['Healthy life expectancy'].mean(),linestyle='--',color='black')
ax[3].set_title('Healthy life expectancy',fontsize=30)
sns.barplot(y=reg_df['Freedom to make life choices'].mean().values,x=reg_df['Freedom to make life choices'].mean().index,ax=ax[4])
ax[4].axhline(df['Freedom to make life choices'].mean(),linestyle='--',color='black')
ax[4].set_title('Freedom to make life choices',fontsize=30)
sns.barplot(y=reg_df['Generosity'].mean().values,x=reg_df['Generosity'].mean().index,ax=ax[5])
ax[5].axhline(df['Generosity'].mean(),linestyle='--',color='black')
ax[5].set_title('Generosity',fontsize=30)
sns.barplot(y=reg_df['Perceptions of corruption'].mean().values,x=reg_df['Perceptions of corruption'].mean().index,ax=ax[6])
ax[6].axhline(df['Perceptions of corruption'].mean(),linestyle='--',color='black',label='Avg line')
ax[6].set_title('Perceptions of corruption',fontsize=30)

plt.xticks(rotation=90,fontsize=25)
plt.legend(bbox_to_anchor=(1.05,1),fontsize=15)
plt.tight_layout()
plt.show()
```

Remember that we have divided the data into 4 parts based on the Ladder score. Let's see how that is distributed among regions.

```python
plt.figure(figsize=(12,10))
plt.style.use('fivethirtyeight')
sns.countplot(y='Regional indicator',hue='Quartile',data=df)
plt.tight_layout()
plt.legend()
plt.show()
```

Western Europe has the maximum number of the countries belonging to the First part. Same as this, Sub-Saharan Africa region has the maximum number of the countries belonging to the Forth part. This show that some region have the group of the approximately same ladder score countries. These region require special attention of the globel development communities.

From the above summery, we can see that the mean and median is almost same for all the columns. now let's see how they are correlated to each other.

```python
fig ,ax =plt.subplots(nrows=3,ncols=2,figsize=(15,12),sharex=True)
plt.style.use('fivethirtyeight')
fig.suptitle('Correlation between Ladder score and other parameter',fontsize=30)
sns.scatterplot(y='Logged GDP per capita',x='Ladder score',hue='Quartile',data=df,ax=ax[0,0])
sns.scatterplot(y='Social support',x='Ladder score',hue='Quartile',data=df,ax=ax[0,1])
sns.scatterplot(y='Healthy life expectancy',x='Ladder score',hue='Quartile',data=df,ax=ax[1,0])
sns.scatterplot(y='Freedom to make life choices',x='Ladder score',hue='Quartile',data=df,ax=ax[1,1])
sns.scatterplot(y='Generosity',x='Ladder score',hue='Quartile',data=df,ax=ax[2,0])
sns.scatterplot(y='Perceptions of corruption',x='Ladder score',hue='Quartile',data=df,ax=ax[2,1])
plt.tight_layout()
plt.legend()
plt.show()
```

```python
corr=df.drop(columns='Ladder score in Dystopia').corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True,cmap=plt.cm.autumn,linewidth=1,linecolor='White',vmin=-1,vmax=1)
plt.show()
```

*Ladder score*, *Logged GDP per capita*, *Social support* and *Healthy life expectancy* are correlated to each other. *Generosity* is not relating to any other parameter. It may be understood that in the developed countries people need help very little, but what surprising is that low ladder scoring countries also not showing any type of relation between *generosity* and other parameter.

```python
SA_data=df.loc[df['Regional indicator']=='South Asia'][:]
feature=['Ladder score',
       'Logged GDP per capita', 'Social support', 'Healthy life expectancy',
       'Freedom to make life choices', 'Generosity',
       'Perceptions of corruption']
plt.figure(figsize=(15,12))
for a in feature:
    plt.subplot(4,2,feature.index(a)+1)
    sns.barplot(x=SA_data.drop(index=SA_data[SA_data['Country name']=='India'].index)['Country name'],y=SA_data[a])
    plt.subplot(4,2,feature.index(a)+1).axhline(SA_data.loc[SA_data['Country name']=='India'][a].values,ls='dotted',color='black',label="India's score")

plt.tight_layout()
plt.legend(bbox_to_anchor=(1.05,1))
plt.show()
```
