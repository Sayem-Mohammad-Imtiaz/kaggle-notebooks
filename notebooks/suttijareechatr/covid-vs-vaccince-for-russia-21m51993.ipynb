# **However, we want to analyze only one country. So in this case I will do russia.**

```python
Selected_country_1 = 'Russia'
Selected_country_2 = 'Russia'
df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1 = df1[df1['country']==Selected_country_1]
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2 = df2[df2['country']==Selected_country_2]
df1.index = pd.to_datetime(df1['date'])
df2.index = pd.to_datetime(df2['date'])

fig,ax=plt.subplots(figsize=(10,5))

ax.plot(df1.index,df1['daily_new_cases'],'b',label='Cases')
ax1 = ax.twinx()
ax1.plot(df2.index,df2['daily_vaccinations'],'r',label='Vaccinations')
plt.show()
```

# To improve the visualization 

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_new_cases = go.Scatter(x=df1.index,y= df1['daily_new_cases'],yaxis='y1',name='Daily confirm cases')
daily_vaccinations = go.Scatter(x=df2.index,y= df2['daily_vaccinations'],yaxis='y2',name='Daily vaccinations')

layout_obj = go.Layout(title='COVID vs Vaccination in Russia',xaxis=dict(title='Date'),yaxis=dict(title='Daily cases'),yaxis2=dict(title='Vaccinations',side='right',overlaying='y'))
fig = go.Figure(data=[daily_new_cases,daily_vaccinations],layout=layout_obj)
iplot(fig)
```

From the upper graph we set the red line to be the number of daily vaccinations and the blue line is the number of daily new case in Russia. the red line will refer to the scale in the right handside. The bluw line will refer to the scale in the left handside. The data show that the vaccinations have help to reduce the number of daily case. We can saw that the number of infected is jump to the moon around September,2020. It's due to the problem that Russia goverment reduce their restict and allow the commercial flights to many region in August,2020 from the reducing of the daily infacted.

# Chack the global ranking for total number of infected cases

```python
df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1 = df1.groupby('country').sum()['daily_new_cases'].sort_values(ascending=False)[0:10]
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2 = df2.groupby('country').max()['people_fully_vaccinated_per_hundred'].sort_values(ascending=False)[0:90]
print(df1)
print(df2.to_string())
```

So we can saw that Russia is in the top 10 of the global ranking for total number of infected cases. This country was ranking as 6th rank from the top. However, the ranking of Russia for total vaccinations per number of people is 85th from max to min.
