# Create a global map for covid-19 Vaccinations

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import plotly.express as px
import plotly

import os

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df.index = pd.to_datetime(df['date'])

fig = px.choropleth(df,locations='country',locationmode='country names',color='people_vaccinated_per_hundred',hover_name='country',animation_frame='date',color_continuous_scale='GnBu',range_color=(0.0,70.0))
fig.update_layout(title_text='Percentage of people vaccinated in each country',title_x=0.5)
fig.show()
```

I choose to plot the data in the column that contain the percentage of people vaccinated in each country to see how much people in each country get the vaccine per day. I choose to use the color in band of blue and green to make it easy to read. In my opinion, to use too many color to make a graph will cause confusion

# Earthqueak record in our World

```python
df = pd.read_csv('../input/202106-recent-e/query.csv',header=0)
df.index = pd.to_datetime(df['time'])
print(df.columns)
df['time'] = df.index.strftime('%Y-%m-%d %H:00:00')
df.sort_index()
fig = px.scatter_geo(df,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='rainbow',range_color=(1.0,5.0))
fig.show()
```
