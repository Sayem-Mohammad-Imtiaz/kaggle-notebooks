## <center> Creating Models </center>

```python

x = df.iloc[:,0:4] #feature values
y = df.iloc[:,4:] #target
```

```python

x.head(3)
```

```python
y.head(3)
```

```python
# Splitting the Data Set into Independent Variables and Dependent Variables

from sklearn.model_selection import train_test_split

x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=8)
```

## <center> Classification Algorithms </center>

## <center> Decision Tree </center>

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

clf = DecisionTreeClassifier(criterion = 'entropy')
model=clf.fit(x_train,y_train)
prob=clf.predict_proba([[4,3.2,5,6]])

y_pred = clf.predict(x_test)
accuracy_score_clf = accuracy_score(y_test,y_pred,3)

#y_pred = clf.predict(x_test)

#acc_1a = metrics.accuracy_score(y_test,y_pred,3)
#pre_1a = metrics.precision_score(y_test,y_pred)
#print("Accuracy:", acc_1a, pre_1a)

print(accuracy_score_clf)
print(prob)

cm_dtc = confusion_matrix(y_test,y_pred)
```

```python
#pretty table
from prettytable import PrettyTable

AccuracyTable = PrettyTable(["Algorithm", "Accuracy "])
AccuracyTable.add_row(["Decision Tree", accuracy_score_clf])
print(AccuracyTable)
```

## <center> Random Forest </center>

```python

from sklearn.ensemble import RandomForestClassifier #Importing lib
clf = RandomForestClassifier(n_estimators = 9, criterion = 'entropy') #creating a classifier 
model=clf.fit(x_train,y_train) #training the model using training sets
y_pred = clf.predict(x_test)

cm_rfc = confusion_matrix(y_test,y_pred)
```

Calculate Accuracy of the model

```python
from sklearn import metrics

acc=metrics.accuracy_score(y_test,y_pred,3)
print(acc)
```
