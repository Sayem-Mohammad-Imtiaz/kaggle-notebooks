```python
data['Total Score']=data['math score']+data['reading score']+data['writing score']
```

```python
def result(TS,MS,WS,RS ):
    if(TS>120 and MS>40 and WS>40 and RS>40):
        return 'P'
    else:
        return 'F'
```

```python
data['Pass/Fail']=data.apply(lambda x: result(x['Total Score'],x['math score'],x['writing score'],x['reading score']),axis = 1 )
```

```python
data.head(20)
```

```python
plt.pie(data['Pass/Fail'].value_counts(),labels=['Pass','Fail'],autopct='%1.1f%%')
plt.title('Percentage of students Passed/Failed')
```

```python
#Average Math score and Math level for average
threshold = sum(data['math score'])/len(data['math score'])
print("threshold",threshold)
data["math_level"] = ["good" if i > threshold else "bad" for i in data['math score']]
data.loc[:20,["math_level","math score"]]
```
