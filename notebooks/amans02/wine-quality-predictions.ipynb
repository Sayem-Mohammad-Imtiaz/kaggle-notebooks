```python
df=pd.read_csv('../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
```

```python
df.head()
```

```python
#coverting quality into binary thaking 6 threshold
df['quality']=df['quality'].apply(lambda x : 1 if x>6 else 0)
df.head()
```

```python
#checkimg corelation
corr=df.corr()
sns.heatmap(corr)
s = corr.unstack()
sorting_corr = s.sort_values(kind="quicksort")
```

```python
print(sorting_corr[(sorting_corr>0.6) & (sorting_corr!=1)])
```

```python
#importing librabries, I add the common imports in this cell
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, roc_curve, f1_score, accuracy_score, recall_score
from sklearn.feature_selection import VarianceThreshold
```

```python
#selection using variance threshold
X=df.drop('quality',axis=1)
y=df['quality']
sel = VarianceThreshold(threshold=(.8 * (1 - .8)))
sel.fit(X)
X=X[X.columns[sel.get_support(indices=True)]]
print(X)
#split train test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,shuffle=False)
```

# Logistic Regrression

```python
#prediction with Logistic regrression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
pred=lr.predict(X_test)
print(confusion_matrix(y_test,pred))
print("Accuracy :",accuracy_score(y_test,pred),"\n f1 acore :",f1_score(y_test,pred),"\n recall :",recall_score(y_test,pred))
fpr, tpr, threshold = roc_curve(y_test, pred)
print(fpr,tpr)
plt.plot(fpr,tpr,'b')
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
print(confusion_matrix(y_test,pred))
print("Accuracy :",accuracy_score(y_test,pred),"\n f1 acore :",f1_score(y_test,pred),"\n recall :",recall_score(y_test,pred))
fpr, tpr, threshold = roc_curve(y_test, pred)
print(fpr,tpr)
plt.plot(fpr,tpr,'b')
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
pred=knn.predict(X_test)
print(confusion_matrix(y_test,pred))
print("Accuracy :",accuracy_score(y_test,pred),"\n f1 acore :",f1_score(y_test,pred),"\n recall :",recall_score(y_test,pred))
fpr, tpr, threshold = roc_curve(y_test, pred)
print(fpr,tpr)
plt.plot(fpr,tpr,'b')
```
