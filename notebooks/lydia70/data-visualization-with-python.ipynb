```python

#Scatterplot Clusters

#select data
df_score = df[["Annual Income (k$)", "Spending Score (1-100)"]]

#import the K-means package
import numpy as np
from sklearn.cluster import KMeans


cluster_maximum = []

prediction = df_score.values

for i in range(1,15):
    kmeans = KMeans(n_clusters = i, init='k-means++', n_init=10,  max_iter = 300, algorithm='elkan')
    kmeans.fit_predict(prediction)
    cluster_maximum.append(kmeans.inertia_)

plt.plot(np.arange(1, 15), cluster_maximum)
plt.xlabel("Number")
plt.ylabel("Sum of Squares")
plt.show()

#elbow of the scree plot is at number 5, number 5 is the optimal number of clusters
for i in range(1,6):
    kmeans = KMeans(n_clusters = i,init='k-means++', n_init=10,  max_iter = 300, algorithm='elkan')
    kmeans.fit_predict(prediction)
    cluster_maximum.append(kmeans.inertia_)

# Compute cluster centers
cluster = kmeans.predict(prediction)



#make scatterplot
plt.figure(figsize=(12,8))
plt.scatter(prediction[cluster == 0, 0], prediction[cluster == 0, 1], c = 'blue', label = 'Cluster 1')
plt.scatter(prediction[cluster == 1, 0], prediction[cluster == 1, 1], c = 'green', label = 'Cluster 2')
plt.scatter(prediction[cluster == 2, 0], prediction[cluster == 2, 1], c = 'black', label = 'Cluster 3')
plt.scatter(prediction[cluster == 3, 0], prediction[cluster == 3, 1], c = 'yellow', label = 'Cluster 4')
plt.scatter(prediction[cluster == 4, 0], prediction[cluster == 4, 1], c = 'red', label = 'Cluster 5')

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:, 1],s = 300, c = 'orange', label = 'Centroids', marker='*')

plt.title('Mall Customer Clusters')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend(loc='upper right')

plt.show()
```
