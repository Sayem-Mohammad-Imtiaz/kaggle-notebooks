```python
#encoding transmission and fueltype column to numeric

data['transmission'] = data['transmission'].map({'Manual':1,'Automatic':2,'Semi-Auto':3})
```

```python
data['fuelType'] = data['fuelType'].map({'Petrol':1,'Diesel':2})
```

```python
data['fuelType'].fillna(data['fuelType'].mode()[0],inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
data['model'] = le.fit_transform(data['model'])
```

```python
data.head()
```

```python
X= data.drop('price',axis=1)
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
coeff_df = pd.DataFrame(lr.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
from sklearn import metrics
```

```python
predictions = lr.predict(X_test)
```

```python
plt.scatter(y_test,predictions,color='purple')
```

```python
sns.distplot((y_test-predictions),bins=50,color='green')
```

```python
print('MAE : ',metrics.mean_absolute_error(y_test,predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
