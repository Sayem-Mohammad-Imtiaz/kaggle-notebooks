```python
# check the data type for each column in the data-frame 
the_data.dtypes
```

We can see that Date column is object, next we will convert it to a DateTime.

```python
# check the data type of the date column
# Date is object not date, so we will convert it to date 
# To do so, we will create new coulmn, name it as 'p_date' and copy all the
# date in 'Date' to the new coulmn then convert it to Date dtype.



# Copy first 10 charecters from Date column to p_date column 
the_data["p_date"] = the_data['Date'].apply(lambda x: x[0:10]) 


# convert p_date dtype to date.
the_data['p_date'] = pd.to_datetime(the_data['p_date'])
```

```python
# check the dtype for the p_date
print(the_data['p_date'].dtypes)
```
