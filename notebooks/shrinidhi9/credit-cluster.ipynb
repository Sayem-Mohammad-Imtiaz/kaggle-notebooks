```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(context="notebook", palette="Spectral", style = 'darkgrid' ,font_scale = 1, color_codes=True)
```

```python
dataset = pd.read_csv('../input/CC GENERAL.csv')
```

```python
# import pandas_profiling
# profile = pandas_profiling.ProfileReport(dataset)
# profile
```

```python
dataset.drop('CUST_ID', axis=1, inplace=True) 
dataset.drop('PURCHASES', axis=1, inplace=True) # high correlaion
```

```python
print(dataset.columns[dataset.isnull().any()])
print(dataset.isnull().sum())
```

```python
dataset['MINIMUM_PAYMENTS'] = dataset['MINIMUM_PAYMENTS'].fillna(dataset['MINIMUM_PAYMENTS'].median())
dataset['CREDIT_LIMIT'] = dataset['CREDIT_LIMIT'].fillna(dataset['CREDIT_LIMIT'].median())
```

```python
#Duplicate check
print(sum(dataset.duplicated(dataset.columns)))
```

```python
d_s = dataset.skew()[abs(dataset.skew()) >= 2].index
d_s
```

```python
for col in d_s:
    dataset[col] = np.cbrt(dataset[col])
```

```python
#Remove outliers - 
from scipy import stats
numeric_cols = dataset.select_dtypes(include=['int64','float64'])
z = np.abs(stats.zscore(numeric_cols))

for i in range(numeric_cols.shape[0]):
    for j in range(numeric_cols.shape[1]):
        if z[i,j] >= 3:
            print(f"{i} -{j}")
            numeric_cols.iloc[i,j] = numeric_cols.iloc[:,j].median()
            
dataset.update(numeric_cols)               
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
dataset = sc.fit_transform(dataset)
```

```python
from sklearn.cluster import KMeans
#elbow method
wcss=[]
for i in range(1,17):
    km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300)
    km.fit_predict(dataset)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
   
plt.plot( wcss, 'ro-', label="WCSS")
plt.title("Computing WCSS for KMeans++")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```
