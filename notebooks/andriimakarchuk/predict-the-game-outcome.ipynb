```python
import pandas as pd
from sklearn import metrics
```

```python
data = pd.read_csv("../input/league-of-legends-diamond-ranked-games-10-min/high_diamond_ranked_10min.csv").dropna()
data.head()
```

```python
X = data.drop( labels=["blueWins"], axis=1 )
y = data["blueWins"]
```

```python
X.corr()
```

```python
corrTable = X.corr()
cols = X.columns
```

```python
newCols = []
for i in cols:
    for j in cols:
        if( i!=j and abs(corrTable[i][j])<0.01 ):
            newCols.append(j)
newCols = list( pd.Series(newCols).unique() )
print( len(newCols) )
X = X[newCols]
```

Ok. Data is mpdofied as we need. Let's build two classification models to predict a win of blue command.

```python
from sklearn import tree
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
treeClassifier = tree.DecisionTreeClassifier(
    min_samples_leaf=100
).fit(X_train, y_train)
```

```python
print( "Accuracy for tree classifier: "+str( metrics.accuracy_score(y_test, treeClassifier.predict(X_test)) ) )
```

```python
tree.plot_tree(treeClassifier)
```

```python
from sklearn.linear_model import LogisticRegression
logRegModel = LogisticRegression()
logRegModel.fit(X_train, y_train)
print( "Accuracy for logistic regressiom model: "+str( metrics.accuracy_score(y_test, logRegModel.predict(X_test)) ) )
```

As we see, **classification tree is good model for predicting wining of blue command**.
