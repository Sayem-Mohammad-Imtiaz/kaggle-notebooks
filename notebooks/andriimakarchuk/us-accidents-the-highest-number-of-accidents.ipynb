**Task: which US state has the highest number of accidents, and a description of the accidents that usually occur in that state.**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Firstly, **let's read data**. We won't delete objects with missed parameters because it can change an order of list with numbers of accidents.

```python
data = pd.read_csv("../input/us-accidents/US_Accidents_June20.csv")
data.head()
```

```python
print(data.columns)
```

Ok, data is read. Now **let's build rating list of numbers of accidents for states**.

```python
numberOfAccidents = data[["ID", "State"]].groupby("State").count().sort_values("ID", ascending=False)
```

In the beginning, let's show 10 states with highest numbers of accidents.

```python
numberOfAccidents.head(10)
```

Now let's visualize full list.

```python
sns.barplot(numberOfAccidents["ID"], numberOfAccidents.index)
```

**As we see, the biggest number of accidents have next states: California, Texas and Florida.**
