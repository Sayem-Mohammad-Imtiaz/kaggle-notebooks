```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading and cleaning data.

```python
dataList = [
    pd.read_csv("../input/world-happiness/2015.csv").dropna().drop(["Region", "Happiness Rank"], axis=1).groupby("Country").mean(),
    pd.read_csv("../input/world-happiness/2016.csv").dropna().drop(["Region", "Happiness Rank"], axis=1).groupby("Country").mean(),
    pd.read_csv("../input/world-happiness/2017.csv").dropna().drop("Happiness.Rank", axis=1),
    pd.read_csv("../input/world-happiness/2018.csv").dropna().drop("Overall rank", axis=1),
    pd.read_csv("../input/world-happiness/2019.csv").dropna().drop("Overall rank", axis=1)
]
```

Firstly, we need to process data as we need. Let's do it.

```python
for data in dataList:
    for column in data.columns:
        if(data[column].dtype=="object"):
            values = list( data[column].unique() )
            for value in values:
                data[column][ data[column]==value ] = values.index(value)
```

```python
for data in dataList:
    data.astype("float")
```

To see how happiness changed with time let's claster every data package using mark from 0 (lowest) to 4 (highest) to mark happiness on countries for every year.

```python
from sklearn.cluster import KMeans
```

```python
clustersDataPackage = [ [2015+i, pd.Series(KMeans(5).fit_predict(dataList[i]))] for i in range(len(dataList)) ]
```

```python
sns.countplot( clustersDataPackage[0][1] )
```
