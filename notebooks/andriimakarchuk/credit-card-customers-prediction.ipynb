# Data preprocessing
To build prediction model **we should read data and preprocess it how we need**. Let's do it.

```python
data = pd.read_csv("../input/credit-card-customers/BankChurners.csv").dropna()
data.head()
```

```python
data.info()
```

Let's rewrite categorial data in numeric variant.

```python
Attrition_Flags = list(data["Attrition_Flag"].unique())
for flag in Attrition_Flags:
    print(flag)
```

```python
for flag in Attrition_Flags:
    data["Attrition_Flag"][ data["Attrition_Flag"]==flag ] = Attrition_Flags.index(flag)
```

```python
Genders = list(data["Gender"].unique())
for gender in Genders:
    print(gender)
```

```python
for gender in Genders:
    data["Gender"][ data["Gender"]==gender ] = Genders.index(gender)
```

```python
Education_Levels = list(data["Education_Level"].unique())
for level in Education_Levels:
    print(level)
```

```python
for level in Education_Levels:
    data["Education_Level"][ data["Education_Level"]==level ] = Education_Levels.index(level)
```

```python
Marital_Statuses = list(data["Marital_Status"].unique())
for status in Marital_Statuses:
    print(status)
```

```python
for status in Marital_Statuses:
    data["Marital_Status"][ data["Marital_Status"]==status ] = Marital_Statuses.index(status)
```

```python
Income_Categories = list(data["Income_Category"].unique())
for category in Income_Categories:
    print(category)
```

```python
for category in Income_Categories:
    data["Income_Category"][ data["Income_Category"]==category ] = Income_Categories.index(category)
```

```python
Card_Categories = list(data["Card_Category"].unique())
for category in Card_Categories:
    print(category)
```

```python
for category in Card_Categories:
    data["Card_Category"][ data["Card_Category"]==category ] = Card_Categories.index(category)
```

```python
data.astype("float")
data.head()
```

# Data clustering

Ok, data is processed. Let's cluster it.

```python
from sklearn.cluster import KMeans
```

```python
X = data
y = list(KMeans(n_clusters=2).fit(X).labels_)
```

# Building classification model

Data is clustered. Let's build classifier. We will build and test logistic regression model and model, built on decision tree.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn import tree
from sklearn import metrics
import seaborn as sns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

Firstly, let's build and test logistic regression model.

```python
logPredModelDeg1 = LogisticRegression().fit(X_train, y_train)
```

Balanced accuracy.

```python
print(metrics.balanced_accuracy_score(y_test, logPredModelDeg1.predict(X_test)))
```

Test f1.

```python
print(metrics.f1_score(y_test, logPredModelDeg1.predict(X_test)))
```

As we see, built model is bad. Les's try to modify it.

```python
logPredModelDeg2 = LogisticRegression().fit(PolynomialFeatures(degree=2).fit_transform(X_train), y_train)
```

Bakanced accuracy.

```python
print(metrics.balanced_accuracy_score(y_test, logPredModelDeg2.predict(PolynomialFeatures(degree=2).fit_transform(X_test))))
```

Test f1.

```python
print(metrics.f1_score(y_test, logPredModelDeg2.predict((PolynomialFeatures(degree=2).fit_transform(X_test)))))
```

As we see, models based in logistic regression are not good in this situation. Sometimes we can have better results using decision trees. Let's try to build prediction model based on decision tree.

```python
treeClassifierDeg1 = tree.DecisionTreeClassifier(min_samples_leaf=250).fit(X_train, y_train)
```

```python
Balansed accuracy.
```

```python
print(metrics.balanced_accuracy_score(y_test, treeClassifierDeg1.predict(X_test)))
```

```python
Test f1.
```

```python
print(metrics.f1_score(y_test, treeClassifierDeg1.predict(X_test)))
```

ROC AUC.

```python
print(metrics.roc_auc_score(y_test, treeClassifierDeg1.predict(X_test)))
```

As we see using some metrics, model based decision tree is very good. It's a reason to try predict churned customers using this instrument.
