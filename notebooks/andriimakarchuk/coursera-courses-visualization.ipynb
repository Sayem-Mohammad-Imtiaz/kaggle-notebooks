**Task: visualize the dataset by complete analysis on it.**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../input/coursera-course-dataset/coursea_data.csv").dropna()
data.head()
```

# Data clearning and modifying
To do some visualizations, firstly we need to select parameters, which we can visualize, and modify it as we need.

```python
data = data[ ["course_organization", "course_Certificate_type", "course_rating", "course_difficulty", "course_students_enrolled"] ]
data.head()
```

```python
data.info()
```

```python
organizations     = data["course_organization"].value_counts().sort_values(ascending=False)[0:3]

enrolled = data["course_students_enrolled"].to_list()
enrolledSums = [ float(enr.split("k")[0].split("m")[0]) for enr in enrolled ]
```

# Visualizing
Data is transformated as we need. Now we can make visualizations. Firstly, let's show TOP-3 of organizations by numbers of published cources.

```python
plt.plot(organizations.index, organizations)
```

Now let's show counts of evety certificate type.

```python
sns.countplot(data["course_Certificate_type"])
```

Now let's see estimated counts of courses per every difficulty level.

```python
sns.countplot(data["course_difficulty"])
```

Now let's see a "behaviour" or rating.

```python
sns.histplot(data["course_rating"])
```

And, on the end, let's see "behaviour" of distribution of enrolled sudents.

```python
sns.histplot(enrolledSums)
```
