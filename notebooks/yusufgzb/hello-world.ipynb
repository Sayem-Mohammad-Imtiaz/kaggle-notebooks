```python
data=pd.read_csv("/kaggle/input/students-performance-in-exams/StudentsPerformance.csv")

data.info()
```

**filtering**

```python
filtre1=data.loc[:,["gender","math score","reading score","writing score"]]
filtre1
```

```python
filtreF=data[(data["math score"]>80)&(data["reading score"]>80)&(data["gender"]=="female")&(data["writing score"]>80)][["gender","lunch","math score","reading score","writing score"]]
filtreM=data[(data["math score"]>80)&(data["reading score"]>80)&(data["gender"]=="male")&(data["lunch"]=="standard")&(data["writing score"]>80)][["gender","lunch","math score","reading score","writing score"]]

filtreM
```

```python
filtreF.plot(kind='hist',y="math score",bins=50,range=(80,100),normed=True)
filtreF.plot(kind='hist',y="reading score",bins=50,range=(80,100),normed=True)
filtreF.plot(kind='hist',y="writing score",bins=50,range=(80,100),normed=True)

plt.show()
```
