```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler,MinMaxScaler

from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split,cross_val_score,StratifiedKFold
from sklearn.feature_selection import SelectKBest,chi2,mutual_info_classif
from sklearn.ensemble import GradientBoostingClassifier,RandomForestClassifier
```

```python
train=pd.read_csv('/kaggle/input/health-care-data-set-on-heart-attack-possibility/heart.csv')
```

```python
scale=MinMaxScaler()
train[['age','trestbps','chol','thalach']]=scale.fit_transform(train[['age','trestbps','chol','thalach']])
```

```python
plt.figure(figsize=(15,9))
sns.heatmap(train.corr('kendall'),annot=True)
```

```python
xtrain=train.drop('target',axis=1)
ytrain=train['target']

pipeline=Pipeline(steps=[('select',SelectKBest(score_func=mutual_info_classif,k=7)),('model',GradientBoostingClassifier())])
score=cross_val_score(pipeline,xtrain,ytrain,cv=5,scoring='accuracy')
print(np.mean(score))
```
