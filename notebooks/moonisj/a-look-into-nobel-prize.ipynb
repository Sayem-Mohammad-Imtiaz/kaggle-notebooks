```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
nobel = pd.read_csv('../input/nobel-laureates/archive.csv')
```

```python
nobel.head()
```

```python
by_gender = nobel['Sex'].value_counts().reset_index()
```

```python
fig = px.pie(by_gender, values='Sex', names='index', title = 'Female vs. Male Winners of the Nobel Price since 1901')
fig.show()
```

**List of All Female Nobel Winners**

```python
female_winners = nobel[nobel['Sex'] == 'Female']
female_winners = female_winners[['Year','Category','Motivation', 'Full Name']]
female_winners
```

```python
nobel['Category'].unique()
```

```python
che_cat = nobel[nobel['Category'] == 'Chemistry']
lit_cat = nobel[nobel['Category'] == 'Literature']
med_cat = nobel[nobel['Category'] == 'Medicine']
pea_cat = nobel[nobel['Category'] == 'Peace']
phy_cat = nobel[nobel['Category'] == 'Physics']
eco_cat = nobel[nobel['Category'] == 'Economics']
```

```python
che_gender = che_cat['Sex'].value_counts().reset_index()
lit_gender = lit_cat['Sex'].value_counts().reset_index()
med_gender = med_cat['Sex'].value_counts().reset_index()
pea_gender = pea_cat['Sex'].value_counts().reset_index()
phy_gender = phy_cat['Sex'].value_counts().reset_index()
eco_gender = eco_cat['Sex'].value_counts().reset_index()
```

```python
fig=go.Figure()
fig.add_trace(go.Bar(x=che_gender['index'],y=che_gender['Sex'], name='Chemistry' ))
fig.add_trace(go.Bar(x=lit_gender['index'],y=lit_gender['Sex'], name='Literature'))
fig.add_trace(go.Bar(x=med_gender['index'],y=med_gender['Sex'], name='Medicine'))
fig.add_trace(go.Bar(x=pea_gender['index'],y=pea_gender['Sex'], name='Peace'))
fig.add_trace(go.Bar(x=phy_gender['index'],y=phy_gender['Sex'], name='Physics'))
fig.add_trace(go.Bar(x=eco_gender['index'],y=eco_gender['Sex'], name='Economics'))

fig.update_layout(barmode='group', title = 'Category Wise: Women vs. Men Nobel Prize Winners since 1901')
fig.show()
```

```python
top_10 = nobel['Birth Country'].value_counts().reset_index()
top_10 = top_10.sort_values('Birth Country', ascending=False).head(10)
```

```python
fig = px.bar(top_10, x='index',y='Birth Country', title = 'Top 10 Nobel Winning Countries')
fig.show()
```

```python
top_10_org = nobel['Organization Name'].value_counts().reset_index()
top_10_org = top_10_org.sort_values('Organization Name', ascending=False).head(10)
```

```python
fig = px.bar(top_10_org, x='index',y='Organization Name', title = 'Organizations Producing Most Nobel Laureats')
fig.show()
```

**Youngest ever Nobel Winners**

```python
nobel['Birth Date'] = nobel['Birth Date'].replace({'1898-00-00':'1898-01-01'})
nobel['Birth Date'] = nobel['Birth Date'].replace({'1943-00-00':'1943-11-07'})
nobel['Birth Date'] = pd.to_datetime(nobel['Birth Date'])
nobel['year'] = pd.DatetimeIndex(nobel['Birth Date']).year
nobel['age'] = nobel['Year'] - nobel['year']
young = nobel[['Full Name','Category', 'age']]
young = young.sort_values('age').head(10)
young
```
