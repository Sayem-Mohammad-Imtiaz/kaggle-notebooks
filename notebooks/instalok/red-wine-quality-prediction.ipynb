# **About Dataset**

* **fixed acidity**: most acids involved with wine or fixed or nonvolatile
* **volatile acidity**: the amount of acetic acid in wine
* **citric acid**: found in small quantities, citric acid can add 'freshness' and flavor to wines
* **residual sugar**: the amount of sugar remaining after fermentation stops
* **chlorides**: the amount of salt in the wine
* **free sulfur dioxide**: the free form of SO2 exists in equilibrium between molecular SO2 (as a dissolved gas) and bisulfite ion
* **total sulfur dioxide**: amount of free and bound forms of S02
* **density**: the density of water is close to that of water depending on the percent alcohol and sugar content
* **pH**: describes how acidic or basic a wine is on a scale from 0 (very acidic) to 14 (very basic)
* **sulphates**: a wine additive which can contribute to sulfur dioxide gas (S02) levels
* **alcohol**: the percent alcohol content of the wine
* **quality**: output variable (based on sensory data, score between 0 and 10)

# **Predictions**

Lets change the quality column of dataset to:
* 0 - bad
* 1 - good 

```python
bins = (2, 6.5, 8)
labels = ['bad', 'good']
ds['quality'] = pd.cut(x = ds['quality'], bins = bins, labels = labels)
```

```python
ds['quality'].value_counts()
```

```python
labelencoder_y = LabelEncoder()
ds['quality'] = labelencoder_y.fit_transform(ds['quality'])
```

```python
ds.head()
```

Now our dataset is ready for predictions. Further we will break the datasets into 2 parts:
* x - containing all input columns of original dataset.
* y - containing output column of original dataset.

```python
x = ds.drop(['quality'], axis=1)
y=ds.loc[:,['quality']]
```

```python
x.head()
```

```python
y.head()
```

**Splitting dataset for training and testing.**

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 50)
```

```python
print("Shape of X_train: ",x_train.shape)
print("Shape of X_test: ", x_test.shape)
print("Shape of y_train: ",y_train.shape)
print("Shape of y_test",y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)
```

We will use different algorithms and check which algorithm is giving best accuracy.
* Logistic Regression
* Decision Trees
* Naive Bayes
* Random Forests
* SVM

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score
lr = LogisticRegression()
lr.fit(x_train_scaled, y_train)
lr_predict = lr.predict(x_test_scaled)
```

```python
lr_conf_matrix = confusion_matrix(y_test, lr_predict)
lr_acc_score = accuracy_score(y_test, lr_predict)
print(lr_conf_matrix)
print(lr_acc_score*100)
```

By using Logistic Regression, accuracy is 91.25%.

**Decision Trees**

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train_scaled,y_train)
dt_predict = dt.predict(x_test_scaled)
```

```python
dt_conf_matrix = confusion_matrix(y_test, dt_predict)
dt_acc_score = accuracy_score(y_test, dt_predict)
print(dt_conf_matrix)
print(dt_acc_score*100)
```

By using Decision Tree Classifier, accuracy is 91.5625%.

**Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train_scaled,y_train)
nb_predict=nb.predict(x_test_scaled)
```

```python
nb_conf_matrix = confusion_matrix(y_test, nb_predict)
nb_acc_score = accuracy_score(y_test, nb_predict)
print(nb_conf_matrix)
print(nb_acc_score*100)
```

By using Naive Bayes, accuracy is 90%.

**Random Forests**

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train_scaled, y_train)
rf_predict=rf.predict(x_test_scaled)
```

```python
rf_conf_matrix = confusion_matrix(y_test, rf_predict)
rf_acc_score = accuracy_score(y_test, rf_predict)
print(rf_conf_matrix)
print(rf_acc_score*100)
```

By using Random Forests, accuracy is 96.5625%
