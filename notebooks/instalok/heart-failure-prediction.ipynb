# **About Dataset**

* **age** - age of patient
* **anaemia** - anaemia is a decrease in the total amount of red blood cells (RBCs) or hemoglobin in the blood, or a lowered ability of the blood to carry oxygen. (0-No and 1-Yes)
* **creatinine_phosphokinase** - is an enzyme expressed by various tissues and cell type (mcg/L)
* **ejection_fraction** - is a measurement of the percentage of blood leaving your heart each time it contracts. (percentage)
* **high_blood_pressure** - (0-No and 1-Yes)
* **serum_creatinine** - serum creatinine studies are used to determine renal function.(mg/dL)
* **serum_sodium** - serum sodium is routine in assessing electrolyte, acid-base, and water balance, as well as renal function.(mEq/L)
* **diabetes** - (0-No and 1-Yes)
* **platelets** - are tiny blood cells that help your body form clots to stop bleeding. (kiloplatelets/mL)
* **sex** - (0-Female and 1-Male)
* **smoking** - (0-No and 1-Yes)
* **time** - number of times consulted with doctor (guess)
* **DEATH_EVENT** - (0-No and 1-Yes)

**Prediction using Random Forest Classification**

```python
x = ds.drop(['DEATH_EVENT'], axis=1)
y=ds.loc[:,['DEATH_EVENT']]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.15, random_state=2)
```

```python
clf = RandomForestClassifier(max_depth=3, random_state=3)
clf.fit(x_train, y_train)
pred=clf.predict(x_test)
print("Accuracy of RandomForestClassifier is : ",clf.score(x_test,y_test))
```

# **Confusion Matrix**

```python
titles_options = [("Confusion matrix", None)]
for title, normalize in titles_options:
    disp = plot_confusion_matrix(clf, x_test, y_test,
                                cmap=plt.cm.Blues)
    disp.ax_.set_title(title)

    print(title)
    print(disp.confusion_matrix)

plt.show()
```
