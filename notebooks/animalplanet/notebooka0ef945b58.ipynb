```python
train = pd.read_csv('/kaggle/input/unsw-nb15/UNSW_NB15_training-set.csv')
test = pd.read_csv('/kaggle/input/unsw-nb15/UNSW_NB15_testing-set.csv')

    
```

```python
df1 = pd.DataFrame(train)
df2 = pd.DataFrame(test)
train.head(5)
```

```python
df1.columns
```

```python
features_train = df1[["dpkts", "spkts","label", "attack_cat"]]
final_train = features_train.loc[(df1['attack_cat'] == 'DoS') | (df1['attack_cat'] == 'Normal')]

features_test = df2[["dpkts", "spkts","label", "attack_cat"]]
final_test = features_test.loc[(df2['attack_cat'] == 'DoS') | (df2['attack_cat'] == 'Normal')]


final_train.head(5)
```

```python
final_train['label'].value_counts()
```

```python
X_train = final_train.drop(['label', 'attack_cat'], axis=1)
y_train = final_train['label']

X_test = final_test.drop(['label', 'attack_cat'], axis=1)
y_test = final_test['label']

X_train.shape, X_test.shape
```

```python
print (X_train)
print(X_train.dtypes)
```

```python
# import SVC classifier
from sklearn.svm import SVC


# import metrics to compute accuracy
from sklearn.metrics import accuracy_score


# instantiate classifier with linear kernel and C=1.0
linear_svc=SVC(kernel='rbf', C=1.0) 


# fit classifier to training set
linear_svc.fit(X_train,y_train)


# make predictions on test set
y_pred_test=linear_svc.predict(X_test)


# compute and print accuracy score
print('Model accuracy score with rbf kernel and C=1.0 : {0:0.4f}'. format(accuracy_score(y_test, y_pred_test)))
```
