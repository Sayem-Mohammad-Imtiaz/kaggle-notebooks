```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('../input/student-grade-prediction/student-mat.csv')
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
num_features = [name for name in data.columns if data[name].dtype in ['int64', 'float64']]
cat_features = [name for name in data.columns if data[name].dtype == 'object']
```

```python
cat_features
```

```python
for x in cat_features:
  print(x," = ",data[x].unique())
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
for i in list(cat_features):
    data[i]=le.fit_transform(data[i])
```

```python
data.head()
```

```python
matrix_corr = data.corr()
matrix_corr.G3
```

```python
x = data.drop('G3',axis=1)
y = data['G3']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
weights=reg.coef_
bais=reg.intercept_
print(weights,bais)
```

```python
from sklearn.metrics import r2_score
y_pred = reg.predict(x_test)
print(r2_score(y_test, y_pred))
```

```python
from sklearn import metrics
print('MEAN Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
plt.scatter(y_test,y_pred)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```
