```python
import numpy as np
import pylab as pl
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
import matplotlib.dates as md
from sklearn.utils import shuffle
```

```python
data = pd.read_csv('../input/sf-salaries/Salaries.csv')
data
```

# TotalPay by JobTitle

Here I compare the TotalPay with each Job Title, then grouping the Job Title and get the max, min, and avg of TotalPay

```python
display(data[["JobTitle","TotalPay","Year",]].groupby(["JobTitle","Year"]).agg(["max",'mean',"min"]).style.background_gradient(cmap="CMRmap_r"))
```

# TotalPay with each Job Title (max,mean,min)

```python
data[["JobTitle","TotalPay","Year",]].groupby(["JobTitle","Year"]).agg(["max",'mean',"min"]).plot(kind="line",color =["orange","black","purple"])
plt.title("TotalPay with each Job Title (max,mean,min)", fontsize=10)
plt.ylabel("TotalPay",fontsize=10)
plt.xlabel(" ")
plt.xticks(rotation=90)
plt.show()
```

# Countplot of TotalPay with each Job Title

```python
TP = data.sort_values(by='TotalPay', ascending=False)[:200]
figure = plt.figure(figsize=(10,6))
sns.barplot(y=TP.JobTitle, x=TP.TotalPay,  palette="rocket")
plt.xticks()
plt.xlabel('TotalPay')
plt.ylabel('JobTitle')
plt.title('Countplot of TotalPay with each Job Title')
plt.show()
```

# TotalPayBenefits by JobTitle

Here I compare the TotalPayBenefits with each Job Title, then grouping the Job Title and get the max, min, and avg of TotalPay

```python
display(data[["JobTitle","TotalPayBenefits","Year",]].groupby(["JobTitle","Year"]).agg(["max",'mean',"min"]).style.background_gradient(cmap="GnBu"))
```

# TotalPayBenefits with each Job Title (max,mean,min)

```python
data[["JobTitle","TotalPayBenefits","Year",]].groupby(["JobTitle","Year"]).agg(["max",'mean',"min"]).plot(kind="line",color =["orange","blue","green"])
plt.title("TotalPayBenefits with each Job Title (max,mean,min)", fontsize=20)
plt.ylabel("TotalPayBenefits",fontsize=10)
plt.xlabel(" ")
plt.xticks(rotation=90)
plt.show()
```

# Countplot of TotalPayBenefits with each Job Title

```python
TPB = data.sort_values(by='TotalPayBenefits', ascending=False)[:200]
figure = plt.figure(figsize=(10,6))
sns.barplot(y=TPB.JobTitle, x=TPB.TotalPayBenefits,  palette="viridis")
plt.xticks()
plt.xlabel('TotalPayBenefits')
plt.ylabel('JobTitle')
plt.title('Countplot of TotalPayBenefits with each Job Title')
plt.show()
```
