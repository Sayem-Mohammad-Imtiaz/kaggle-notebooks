```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../input/hr-analytics-job-change-of-data-scientists/aug_train.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
perc = (df.isnull().sum()*100)/len(df)
perc.sort_values(ascending=False)
```

```python
df.company_type = df.company_type.fillna(method='ffill')
```

```python
df.company_size = df.company_size.fillna(method='ffill')
```

```python
df.major_discipline = df.major_discipline.fillna(method='ffill')
```

```python
df.education_level = df.education_level.fillna(method='ffill')
```

```python
df.enrolled_university = df.enrolled_university.fillna(method='ffill')
```

```python
df.experience = df.experience.fillna(method='ffill')
```

```python
perc = (df.isnull().sum()*100)/len(df)
perc.sort_values(ascending=False)
```

```python
df.info()
```

```python
df.head()
```

```python
df.drop(index=[0], inplace=True)
```

```python
column = ['gender', 'relevent_experience', 'enrolled_university', 'education_level']
count=1
for col in column:
    plt.subplot(2,2,count)
    df[col].value_counts().plot.bar()
    fig=plt.gcf()
    fig.set_size_inches(15,5)
    count=count+1   
```

```python
df.major_discipline.unique()
```

```python
df.major_discipline=df.major_discipline.replace('STEM', 'Commerce')
```

```python
df.experience.unique()
```

```python
df.experience = df.experience.str.replace('>','')
```

```python
df.experience = df.experience.str.replace('<','')
```

```python
df.experience = df.experience.astype('int')
```

```python
df.drop('last_new_job', axis=1, inplace=True)
```
