# Importing Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Reading data

```python
df = pd.read_csv('../input/superhero-set/heroes_information.csv')
df.head()
```

```python
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.shape
```

# Data cleaning

```python
df.Height.replace(-99.,99, inplace=True)
```

```python
df.Height.unique()
```

```python
df['Skin color'].replace('-', 'Wheatish', inplace=True)
```

```python
df['Skin color'].unique()
```

```python
df['Alignment'].replace('-', 'Very good', inplace=True)
```

```python
df['Alignment'].unique()
```

```python
df.Weight.replace(-99., 99., inplace=True)
```

```python
df.Weight.unique()
```

It seems data is clean. Let'do some EDA!

# EDA

```python
df.head()
```

```python
column = ['Gender', 'Hair color', 'Eye color', 'Alignment']
count=1
for col in column:
    plt.subplot(4,2, count)
    df[col].value_counts().plot.pie(shadow=True, autopct='%1.1f%%', radius=1.1, textprops={'fontsize':9})
    count=count+1
    plt.subplot(4,2,count)
    plt.tight_layout()
    plt.style.use('fivethirtyeight')
    df[col].value_counts().plot.bar()
    fig=plt.gcf()
    fig.set_size_inches(50,30)
    plt.title('{0}'.format(col))
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    plt.xticks(rotation=45)
    count=count+1
    
```

```python
plt.figure(figsize=(15,8))
fft = ['Race','Skin color']
count=1
for col in fft:
    plt.subplot(2,2,count)
    sns.countplot(x=col, edgecolor='black', alpha=0.7, data=df)
    sns.despine()
    plt.title('Countplot of {}'.format(col))
    plt.xticks(rotation=90)
    count+=1
    
plt.tight_layout()
plt.show()   
```

```python
df_temp = df.set_index('name')
plt.figure(figsize=(12,6))
sns.barplot(x='name', y='Height', data=df, 
            order=df_temp['Height'].sort_values(ascending=False).index[:15])
plt.xticks(rotation=45)
plt.show()
```

```python
df_temp = df.set_index('name')
corr = df_temp.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
plt.figure(figsize=(10,10))
sns.heatmap(corr, mask=mask, center=0, annot=True,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

```python
df_temp = df.set_index('name')
plt.figure(figsize=(12,6))
sns.barplot(x='name', y='Height', data=df,
           order=df_temp['Height'].sort_values(ascending=True).index[:15], color='orange')
sns.barplot(x='name', y='Weight', data=df,
           order=df_temp['Height'].sort_values(ascending=True).index[:15], color='green')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,6))
g = df.groupby('name').sum()['Height'].sort_values(ascending=False).head(15)
g.plot(kind='bar')
plt.xlabel('Name', fontsize=25)
plt.show()
```

```python
num = ['Height', 'Weight']
count=1
for col in num:
    plt.subplot(2,1, count)
    sns.boxenplot(x=df['name'].head(10), y=df[col].head(10), data=df)
    plt.xticks(rotation=45)
    count=count+1
plt.show()
```

```python
plt.subplot(1,3,1)
plt.title('Height')
sns.heatmap(df.pivot_table(columns='Height',index='name',
                               values='Weight'),cbar=False,cmap='coolwarm')
```

```python
pub = df.Publisher.value_counts().to_dict()
pub

plt.figure(figsize=(8,15))
plt.pie(pub.values(), labels=pub.keys(), autopct='%1.1f%%', shadow=True)
plt.show()
```
