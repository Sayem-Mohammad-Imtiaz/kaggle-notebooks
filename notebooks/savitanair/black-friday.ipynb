```python
#Load Libraries
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns #importing seaborn module 
```

```python
train = pd.read_csv('../input/black-friday/train.csv')
test = pd.read_csv('../input/black-friday/test.csv')
```

```python
df = pd.concat([train, test])
df.shape
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years'].value_counts()
```

```python
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].apply(lambda x: x.replace('+', '')
                                if isinstance(x, str) else x).astype(int)
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype('object')
df.Stay_In_Current_City_Years.value_counts()
```

```python
gen = df.Gender.value_counts().to_dict()

plt.figure(figsize=(5,5))
plt.pie(gen.values(), labels=gen.keys(), autopct='%1.1f%%', shadow=True, explode=[0.1,0])
plt.show()

sns.countplot(df.Gender)
```

```python
age = df.Age.value_counts().to_dict()

plt.figure(figsize=(5,5))
plt.pie(age.values(), labels=age.keys(), autopct='%1.1f%%', shadow=True)
plt.show()

sns.countplot(df.Age)
```

```python
occ = df.Occupation.value_counts().to_dict()

plt.figure(figsize=(5,5))
plt.pie(occ.values(), labels=occ.keys(), autopct='%1.1f%%', shadow=True)
plt.show()

sns.countplot(df.Occupation)
```

```python
prod = df.Product_Category_1.value_counts().to_dict()

plt.figure(figsize=(8,8))
plt.pie(prod.values(), labels=prod.keys(), autopct='%1.1f%%', shadow=True)
plt.show()

sns.countplot(df.Product_Category_1)
```

```python
prod2 = df.Product_Category_2.value_counts().to_dict()

plt.figure(figsize=(8,8))
plt.pie(prod2.values(), labels=prod2.keys(), autopct='%1.1f%%', shadow=True)
plt.show()

sns.countplot(df.Product_Category_2.sort_values(ascending=False))
```

```python
sns.distplot(df.User_ID)
```

```python
#variables - Product_ID 
#Lets emove P and convert product id to a integer
df['Product_ID'] = df['Product_ID'].apply(lambda x: x.replace('P', '')
                                if isinstance(x, str) else x).astype(int)

df['Product_ID'] = df['Product_ID'].astype('int')
```

```python
sns.distplot(df.Product_ID, kde=False, color='r')
```

```python
#Variable - Age Vs Purchase
#Catplot Age+Purchase
sns.catplot(x='Age',y='Purchase',kind='point',data=df, order=['0-17', '18-25', '26-35', '36-45',  '46-50', '51-55', '55+'],)
```
