```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../input/hr-analytics-job-change-of-data-scientists/aug_train.csv')
```

```python
df = df.fillna(method='ffill')
```

```python
print(f'Dataset has {len(df.city.unique())} unique cities')
print('*'*20)
print(f'And the top 10 counts are :')
print(df.city.value_counts().head(10))
print('*'*20)

c = df.city.value_counts().head(10)
fig, ax = plt.subplots(1,1,figsize=(12,6))
ax.bar(c.index, c.values, width=0.8, color='y')
```

```python
print(f'Dataset has {len(df.city_development_index.unique())} unique city development index.')
print('*'*20)
print(f'And the top 10 counts are :')
print(df.city_development_index.value_counts().head(10))
print('*'*20)

plt.figure(figsize=(12,6))
sns.distplot(df.city_development_index, bins=50)
```

```python
city_with_dev = df.groupby(df.city).mean()['city_development_index'].reset_index()
city_with_dev.sort_values(by=['city_development_index'], ascending=False).reset_index(drop=True)
city_with_dev.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(df.city.head(15), df.city_development_index.head(15))
```

```python
gen = df.gender.value_counts().to_dict()
gen

fig, ax = plt.subplots(1,1, figsize=(12,6))
ax.pie(gen.values(), labels=gen.keys(), autopct='%1.1f%%')
```
