```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../input/world-happiness/2016.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
reg = df.Region.value_counts()
reg
```

```python
plt.figure(figsize=(12,6))

reg.plot.bar(x=reg.index, y=reg.values, color='red')
plt.title('Top regions', fontsize=25, weight='bold')
plt.xlabel('Regions')
plt.ylabel('Counts')
plt.show()
```

```python
df['Happiness Score'].plot.hist(bins=30)
```

```python
df['Lower Confidence Interval'].plot.box()
```

```python
df_temp = df.set_index('Country')
sns.set(style='darkgrid')
plt.figure(figsize=(12,6))
sns.barplot(x='Country', y='Happiness Score', data=df, 
            order=df_temp['Happiness Score'].sort_values(ascending=False).index[:10], color='orange')
sns.barplot(x='Country', y='Freedom', data=df, 
            order=df_temp['Happiness Score'].sort_values(ascending=False).index[:10], color='red')
plt.xticks(rotation=90)
plt.show()
```
