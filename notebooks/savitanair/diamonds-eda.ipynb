```python
df.cut.unique()
```

```python
cu = df.cut.value_counts()
cu
plt.figure(figsize=(12,6))
sns.barplot(cu.values, cu.index)
plt.xlabel('Number of different types of cuts')
plt.ylabel('Cuts')
plt.title('Cuts value counts')
plt.legend(df.cut)
plt.show()
```

```python
df.color.unique()
```

```python
col = df.color.value_counts()
col

plt.figure(figsize=(12,6))
sns.barplot(col.values, col.index)
plt.xlabel('Values of colors')
plt.ylabel('Different colors')
plt.title('Value counts of colors')
plt.show()
```

```python
df.clarity.unique()
```

```python
cla = df.clarity.value_counts()
cla

plt.figure(figsize=(12,6))
sns.barplot(cla.index, cla.values)
plt.xlabel('Values of clarity')
plt.ylabel('Different clarities')
plt.title('Value counts of clarities')
plt.show()
```

```python
cutt = df.cut.value_counts().to_dict()
cutt

plt.figure(figsize=(8,8))
plt.pie(cutt.values(), labels=cutt.keys(), autopct='%1.1f%%', explode=[0,0,0.1,0.1,0])
plt.show()
```

```python
colo = df.color.value_counts().to_dict()
colo

plt.figure(figsize=(8,8))
plt.pie(colo.values(), labels=colo.keys(), autopct='%1.1f%%', explode=[0,0,0.1,0.1,0,0,0])
plt.show()
```

```python
clar = df.clarity.value_counts().to_dict()
clar

plt.figure(figsize=(8,8))
plt.pie(clar.values(), labels=clar.keys(), autopct='%1.1f%%', explode=[0,0,0.1,0.1,0,0,0,0])
plt.show()
```
