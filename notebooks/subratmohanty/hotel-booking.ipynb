# How EDA can help drive the hotel business

The below exploratory data analysis (EDA) has been performed on a database of hotel booking data in USA. It demonstrates the power of exploratory data analysis. Following key insights are drawn from the analysis-
* 37% of bookings in 2015-17 were cancelled
* Most of the bookings were made within 113 to 214 days prior to checking in. In fact, a booking made more than 214 days before the check-in date and within a year, was 9 times more likely to be cancelled.
* 2016 was the most busy year for the hotel. Bookings tend to be more during winter and also during summer.
* Only 3% of all the bookings were made by previous customers.
* Just 9 European countries contrbute to 74% of all the confirmed bookings.These are the most frequent client countries for the hotel.
* Portuguese,Dutch,Italians,French, Spanish,German and Belgian visitors prefer mostly city hotels whereas Irish and British visitors prefer resort hotels. None of these customers brought children with them during the stay. 
* All the confirmed bookings were made by corporate clients mostly through travel agents, tour operators and direct channels. Portuguese clients booked mostly in groups.
* There is a healthy competition amongst travel agencies in the most frequent client countries.
* It is observed that countries with less bookings tend to spend more lavishly during their stay. However, weak correlation rules out the possibility of linear corelation. 
* In the consumer base of each country, there are extravagant customers.


**Action plan**
* Organization should improve their customer loyalty program to improve customer retention. It should target strategically the customer base in Portugal, Netherlands, Italy, France, Spain, Germany, Belgium.
* For customers booking more than 214 days before the check-in date, other prospective customers need to be put on waiting list in order to ensure hotel rooms stay occupied as the probability of such a customer cancelling the booking is high.
* Hotel should engage more with travel agencies and tour operators while encouraging healthy competion amongst them to increase customer influx. It should consider giving attractive discounts on group bookings to its Portuguese customer base.
* Cause of low number of bookings of customers who spent high amount of money during their stay needs to be investigated - whether it is due to poor satisfaction level or any other factor is influencing their decision

```python
subset_booked=data[data['is_canceled']==0]
subset_canceled=data[data['is_canceled']==1]
print('Number of bookings made =',subset_booked['hotel'].count())
print('Number of cancellations made =',subset_canceled['hotel'].count())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x=data['is_canceled'],y=data['lead_time'])
```

Number of cancellations is nearly half of that of number of confirmed bookings. 50% of the cancellations have been made between 113 days and 214 days before the check-in date. In cases of confirmed bookings, the lead time is more concentrated at around 45 days with fewer such bookings made before 300 days  

```python
total=data[(data['lead_time']>214)&(data['lead_time']<365)]['hotel'].count()
cases=subset_canceled[(subset_canceled['lead_time']>214)&(subset_canceled['lead_time']<365)]['hotel'].count()
print('Probability of booking getting canceled given that lead time is between 214 days and 300 days= ',cases/total)
print('Probability of booking getting canceled= ',cases/data['hotel'].count())
```

```python
x=[]
y=[]
for i in range(400):
    total=data[(data['lead_time']>214)&(data['lead_time']<i)]['hotel'].count()
    cases=subset_canceled[(subset_canceled['lead_time']>214)&(subset_canceled['lead_time']<i)]['hotel'].count()
    x.append(i)
    y.append((cases/total))
plt.scatter(x,y)
plt.title('Probability of booking cancelation vs max. lead time')
#plt.xlabel('Number of days')
#plt.ylabel('Probability of booking getting canceled')
plt.show()
```

The probability of booking getting canceled is 7%. However, if the lead time is between 214 days and 365 days, then it becomes 9 times more likely that the booking will get canceled.
