```python
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob
```

```python
titles_filtered = data.title[data.title.notnull() & ~data.title.str.contains('Comment')]
```

Filter the titles of Myths.

```python
nlp = spacy.load('en_core_web_lg')
spacy_text_blob = SpacyTextBlob()
nlp.add_pipe(spacy_text_blob)

ftitles_docs = []
for i in titles_filtered:
    doc = nlp(i)
    ftitles_docs.append(doc)
```

Find the number of positive and negative titles.

```python
positive = []
negative = []
neutral = []

for i in ftitles_docs:
    if i._.sentiment.polarity < 0:
        negative.append(i)
    elif i._.sentiment.polarity > 0:
        positive.append(i)
    else:
        neutral.append(i)
```

Let's understand the model's perspective on the sentences it has analyzed. Which words led to the different decision?

```python
positive_assess = []
negative_assess = []
neutral_assess = []

assess = []

for i in ftitles_docs:
    if i._.sentiment.polarity < 0:
        for j in i._.sentiment.assessments:
            for k in j[0]:
                negative_assess.append(k)
    elif i._.sentiment.polarity > 0:
        for j in i._.sentiment.assessments:
            for k in j[0]:
                positive_assess.append(k)
    else:
        for j in i._.sentiment.assessments:
            for k in j[0]:
                neutral_assess.append(k)
```
