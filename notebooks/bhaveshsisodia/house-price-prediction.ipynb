```python
X = df.drop(['Price'],axis=1)
Y = df['Price']
```

```python
print(X,Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25,random_state=0)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train,Y_train)
```

```python
pred= model.predict(X_train)
```

```python
print(pred)
```

```python
# R^2 error 
score_1= metrics.r2_score(Y_train,pred)
```

```python
#Mean Absolute error 
score_2 = metrics.mean_absolute_error(Y_train, pred)
```

```python
print("R square error",score_1)
print("Mean squared error",score_2)
```

```python
plt.scatter(Y_train,pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vS Predicted Price")
plt.show()
```

```python
test_data_pred = model.predict(X_test)
```

```python
score_3= metrics.r2_score(Y_test,test_data_pred)
score_4 = metrics.mean_absolute_error(Y_test,test_data_pred)
print("R square error",score_3)
print("Mean squared error",score_4)
```

```python
plt.scatter(Y_test,test_data_pred)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vS Predicted Price")
plt.show()
```
