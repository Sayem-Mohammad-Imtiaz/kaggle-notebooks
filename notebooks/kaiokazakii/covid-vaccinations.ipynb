```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
data=pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv')
```

```python
data = pd.read_csv("/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv")
data.head()
```

COVID Vaccinations 
In my Kaggle notebook I plan to answer the following questions:
What country has the highest amount of people vaccinated.
What vaccine is the most popular and most used.

```python
plt.figure(figsize=(10, 5))
sorted_vaccination_data = data.groupby("country").total_vaccinations.agg("max").sort_values(ascending = False)
sorted_vaccination_data[sorted_vaccination_data> 1300000].plot(kind = "bar")
plt.ylabel("total no of people vaccinated(in 10s of millions)")
plt.title("Countries leading the vaccination race")
```

```python
plt.figure(figsize=(10, 5))
sorted_vaccination_data = data.groupby("country").total_vaccinations.agg("max").sort_values(ascending = False)
sorted_vaccination_data[sorted_vaccination_data> 1300000].plot(kind = "pie")
plt.title("Countries leading the vaccination race")
```

From the infomation in this bar chart and pie chart I can tell  that the U.S. is clearly has the highest amount of people vaccinated. China follows closerly behind the U.S. with over 55 million people vaccinated. 

```python
plt.figure(figsize=(12, 5))
sorted_vaccination_data = data.groupby("vaccines").total_vaccinations.agg("max").sort_values(ascending = False)
sorted_vaccination_data[sorted_vaccination_data> 130000].plot(kind = "bar")
```

```python
plt.figure(figsize=(10, 5))
sorted_vaccination_data = data.groupby("vaccines").total_vaccinations.agg("max").sort_values(ascending = False)
sorted_vaccination_data[sorted_vaccination_data> 1300000].plot(kind = "pie")
```

The bar and pie charts above show the popularity and use of each COVID-19 vaccine.
