```python
air_bnb_df = pd.read_csv('/kaggle/input/us-airbnb-open-data/AB_US_2020.csv')
air_bnb_df.head()
```

```python
air_bnb_df.shape
```

```python
empty_values_df = air_bnb_df.isna().sum().to_frame().reset_index().rename(columns = {'index': 'Columns', 0 : 'Null Count'})
empty_values_df
```

```python
air_bnb_df.corr(method = "kendall")
```

```python
air_bnb_df = air_bnb_df[['neighbourhood','minimum_nights', 'calculated_host_listings_count', 'availability_365', 'price']]
air_bnb_df = air_bnb_df.sort_values(by = 'price', ascending = False)
air_bnb_df.tail()
```

```python
def regplot(x, y, data):
    sns.regplot(x = x, y = y, data = data)
    plt.ylim(0,)
```

```python
regplot('calculated_host_listings_count', 'price', air_bnb_df)
```

```python
regplot('minimum_nights', 'price', air_bnb_df)
```

```python
regplot('availability_365', 'price', air_bnb_df)
```

```python
hood_grp = air_bnb_df.groupby(['neighbourhood']).mean()
hood_grp.sort_values(by = 'price', ascending = True).reset_index().tail()
```
