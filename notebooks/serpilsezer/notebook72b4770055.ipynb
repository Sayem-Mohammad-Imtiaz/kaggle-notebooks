```python
plt.boxplot(suicide.suicides_no.dropna())
```

```python
from scipy.stats import zscore
z_scores=zscore(suicide['suicides/100k pop'].dropna())
for threshold in range(1,5):
    print('threshold value:{}'.format(threshold))
    print('Number of outliers:{}'.format(len(np.where(z_scores>threshold))))
    print('--'*30)
```

```python
x=suicide.groupby(by=['country'])['suicides_no'].sum()
x.sort_values(ascending=False)[:15].plot(kind='bar')
plt.title('Country_suicides_no')
plt.xlabel('Country')
plt.ylabel('Suicides_no')
```

```python
plt.figure(figsize=(15, 4))
plt.subplot(1,2,1)
plt.plot(suicide['country'],suicide['suicides_no'], c = "#FF914D", lw = 3)
plt.subplot(1,2,2)
plt.plot(suicide['country'],suicide['suicides/100k pop'], c = "#22B573", lw = 3)
plt.xticks(rotation = 45, fontsize = 9)
plt.show()
```

```python
suicide_ülke=suicide.groupby(by=['country'])['suicides/100k pop'].mean()
suicide_ülke.sort_values(ascending=False)[:10].plot(kind='bar')
plt.title('100k pop per Country')
plt.xlabel('Country')
plt.ylabel('Suicides/100k pop')
```
