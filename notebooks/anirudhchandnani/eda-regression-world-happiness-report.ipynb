```python
data = pd.read_csv('../input/world-happiness-report-2021/world-happiness-report.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data["Perceptions of corruption"] = data["Perceptions of corruption"].fillna(data["Perceptions of corruption"].mean())
```

```python
data.columns
```

```python
data["Life Ladder"].dtype
```

```python
for i in data.columns:
    if data[i].dtype != "float64" :
        pass
    else:
        data[i] = data[i].fillna(data[i].mean())
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
f, ax = plt.subplots(figsize=(10, 8))
corr = data.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(240,10,as_cmap=True),
            square=True, ax=ax)
```

# Plotting Positively correlated features with Positive affect

```python
f= plt.figure(figsize=(12,5))

#ax=f.add_subplot(121)
sns.relplot(x = "Life Ladder", y = "Positive affect" , hue="Positive affect", palette="ch:r=-.5,l=.75",data = data )
```

```python
f= plt.figure(figsize=(12,5))

#ax=f.add_subplot(121)
sns.relplot(x = "Freedom to make life choices",y = "Positive affect" , hue = "Positive affect" , data = data)
```

## Sorting countries with decreasing Happiness Index

```python
happiness_score = data.groupby('Country name')['Life Ladder'].sum().sort_values(ascending=False).head(35)
happiness_score
```
