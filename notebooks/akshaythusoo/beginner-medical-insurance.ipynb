# Extracting the CSV File

```python
data =  pd.read_csv (directory)

data.info()
```

# Plotting the graph between the Charges and Region

# Can use this too Encoding

 data.loc [ data [ 'sex' ] == 'male' , 'sex' ] = 1
data.loc [ data [ 'sex' ] == 'female' , 'sex' ] = 0

data.loc [ data ['smoker'] == 'yes' , 'smoker' ] = 1
data.loc [ data ['smoker'] == 'no' , 'smoker' ] = 0
 

status = pd.get_dummies ( data[ 'region' ] , drop_first = True )
data = pd.concat ( [data , status ] , axis  = 1 )

del data ['region'] 

# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
data [ 'sex'] = encoder.fit_transform(data['sex'])
data [ 'smoker' ] = encoder.fit_transform ( data [ 'smoker' ])
data [ 'region' ] = encoder.fit_transform ( data [ 'region' ])
```

```python
data.head()
```

```python
X_columns = [ 'age', 'sex', 'bmi', 'children', 'smoker', 'region' ]
Y_columns = [ 'charges' ]
```

```python
df_X = data [ X_columns ]
df_y = data [ Y_columns ]
```

# Splitting the data into Train , Test

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test  = train_test_split ( df_X , df_y , test_size = 0.2 , random_state = 1 )
```

# Linear Regression ( Degree = 1 )

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit ( X_train , y_train )

print( "Train Accuracy = " ,  model.score( X_train , y_train ) * 100 )
print( "Test Accuracy = " ,  model.score( X_test , y_test ) * 100 )
```

# Linear Regression ( Degree = 2 )

```python
 from sklearn.preprocessing import PolynomialFeatures

degree = PolynomialFeatures ( degree = 2 )
df_X2 = degree.fit_transform ( df_X )
df_y2 = degree.fit_transform ( df_y )

X_train , X_test , y_train , y_test  = train_test_split ( df_X2 , df_y2 , test_size = 0.2 , random_state = 1 )
```

```python

twoDegree_model = LinearRegression()
twoDegree_model.fit ( X_train , y_train )

print(" Train Accuracy = " , twoDegree_model.score( X_train , y_train  ) * 100 ) 
print("Test Accuracy = " , twoDegree_model.score( X_test , y_test) * 100 )
```

# Linear Regression ( Degree = 3 )

```python
from sklearn.preprocessing import PolynomialFeatures

degree = PolynomialFeatures ( degree = 3 )
df_X3 = degree.fit_transform ( df_X )
df_y3 = degree.fit_transform ( df_y )

X_train , X_test , y_train , y_test  = train_test_split ( df_X3 , df_y3 , test_size = 0.2 , random_state = 1 )
```

```python
threeDegree_model = LinearRegression()
threeDegree_model.fit ( X_train , y_train )


print(" Train Accuracy = " , threeDegree_model.score ( X_train , y_train ) * 100 ) 
print("Test Accuracy = " , threeDegree_model.score ( X_test , y_test) * 100 )
```

# Linear Regression ( Degree = 4 )

```python
from sklearn.preprocessing import PolynomialFeatures

degree = PolynomialFeatures ( degree = 4 )
df_X4 = degree.fit_transform ( df_X )
df_y4 = degree.fit_transform ( df_y )

X_train , X_test , y_train , y_test  = train_test_split ( df_X4 , df_y4 , test_size = 0.2 , random_state = 1 )
```

```python
fourDegree_model = LinearRegression()
fourDegree_model.fit ( X_train , y_train )

print(" Train Accuracy = " , fourDegree_model.score( X_train , y_train ) * 100 ) 
print("Test Accuracy = " , fourDegree_model.score( X_test , y_test) * 100 )
```
