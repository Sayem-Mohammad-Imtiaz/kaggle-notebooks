```python
insurance_fnl=insurance[insurance['bmi']<47]
```

```python
sns.boxplot(insurance_fnl.bmi,orient='v')
```

```python
sns.boxplot(insurance.children,orient='v')
```

```python
insurance_fnl['smoker']=insurance_fnl.smoker.apply(lambda x: 1 if x=='yes' else 0)
```

```python

insurance_fnl['sex']=insurance_fnl.sex.apply(lambda x: 1 if x=='male' else 0)
```

```python
insurance_fnl.smoker.value_counts()
```

```python
insurance_fnl.region.value_counts()
```

```python
sns.pairplot(insurance_fnl)
```

```python
sns.boxplot(insurance_fnl.charges)
```

```python
insurance_fnl=insurance_fnl[insurance_fnl.charges<=50000]
```

```python
insurance_fnl=pd.concat([insurance_fnl,pd.get_dummies(insurance_fnl['region'],drop_first=True)],axis=1)
```

```python
insurance_fnl=insurance_fnl.drop('region',axis=1)
```

```python
sns.heatmap(insurance_fnl.corr(),annot=True)
```

```python
y=insurance_fnl.pop('charges')
X=insurance_fnl
```

```python

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=100)
```

```python
sns.heatmap(insurance_fnl.corr(),annot=True)
```

```python
lr=LinearRegression()
```

```python
lrm=lr.fit(X_train,y_train)
```

```python
lrm.intercept_
```

```python
lrm.coef_
```

```python
y_predict=lrm.predict(X_test)
```

```python
y_predict
```

```python
r2_score(y_test,y_predict)
```

```python
lrm.score(X_test,y_test)
```
