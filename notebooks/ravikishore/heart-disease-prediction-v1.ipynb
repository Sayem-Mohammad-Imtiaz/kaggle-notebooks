```python
df=pd.read_csv('/kaggle/input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')
```

```python
df.drop('id',inplace=True,axis=1)
```

```python
df.bmi=df.bmi.fillna(df.bmi.mean())
```

```python
gn=pd.get_dummies(df[['gender','ever_married','work_type','Residence_type','smoking_status']])
```

```python
df.drop(['gender','ever_married','work_type','Residence_type','smoking_status'],axis=1,inplace=True)
```

```python
df=pd.concat([df,gn],axis=1)
```

```python
df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
rf=RandomForestClassifier(random_state=42,max_depth=5)
```

```python
Y=df.pop('heart_disease')
X=df
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.7,random_state=42)
```

```python
rf.fit(X_train,Y_train)
```

```python
Y_pred=rf.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn import metrics 
```

```python
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(Y_test, Y_pred))
```
