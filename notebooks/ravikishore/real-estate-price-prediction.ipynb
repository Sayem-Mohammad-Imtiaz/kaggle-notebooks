```python
real_est.drop(['X1 transaction date'],inplace=True,axis=1)
#real_est.drop(['X2 house age'],inplace=True,axis=1)
```

```python
real_est
```

```python
y=real_est.pop('Y house price of unit area')
X=real_est
```

```python
X
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=100)
```

```python
X_test
```

```python
reg = linear_model.LinearRegression()
```

```python
lr=reg.fit(X_train,y_train)
```

```python
lr.coef_
```

```python
y_pred=lr.predict(X_test)
```

```python
rss=y_test-y_pred
```

```python
plt.scatter(rss,y_pred)
```

```python
sns.distplot(rss)
```

```python
r2_score(y_test,y_pred)
```

```python
y_pred_train=lr.predict(X_train)
```

```python
r2_score(y_train,y_pred_train)
```

```python
df = pd.DataFrame({'Actual':y_test,'Predictions':y_pred})
df['Predictions']= round(df['Predictions'])
df.tail()
```
