1. # Diabeties Analysis from BMI

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

**Importing the data and having a brief look at first 5 entries**

```python
data = pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
data.head()
```

# Clean the data

```python
data = data[['BMI', 'Outcome']]
data['BMI'] = data['BMI'].replace(0, np.NaN)
```

```python
print(data.isnull().sum())
```

```python
data = data.dropna()
```

```python
print(data.isnull().sum())
```

```python
data
```

# Train model

```python
feature_column = ["BMI"]
target = ["Outcome"]
```

```python
from sklearn.model_selection import train_test_split

X = data[feature_column]
y = data[target]

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.33, random_state = 7)
```

```python
#from sklearn.linear_model import LogisticRegression
#model = LogisticRegression(n_jobs=-1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
fig, axe = plt.subplots(dpi = 300)
axe.scatter(X_test['BMI'], y_pred, marker='o')
```

```python
from sklearn.metrics import mean_squared_error, r2_score

print('Coefficients: \n', model.coef_)

print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_pred))

print('Coefficient of determination: %.2f'
      % r2_score(y_test, y_pred))
```

```python
y_test_a = [outcome for outcome in y_test['Outcome']]
print(y_test_a)
```

```python
y_pred_a = [1 if o[0] > 0.5 else 0 for o in y_pred]
print(y_pred_a)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test_a, y_pred_a)
```

```python
print('Mean squared error: %.2f'
      % mean_squared_error(y_test_a, y_pred_a))

print('Coefficient of determination: %.2f'
      % r2_score(y_test_a, y_pred_a))
```
