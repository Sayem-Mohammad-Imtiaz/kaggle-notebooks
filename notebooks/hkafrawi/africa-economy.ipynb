I am intrested to see the local currency of North African Countries in USD and compare their trend through out the time frame of the data set.

```python
df['country'].unique()
```

After extracting the countries from the dataset, I can choose the North African Countries which are Egypt, Alegeria, Tunisia and Morocco. Moreover, I have created a new column to change each exchange rate into USD to be able to compare apples to apples. Diving the rate given by 1 will give us the price of their local currency in USD.

```python
df['exch_ccy_usd'] = 1/df['exch_usd']
df_egy = df[df['country'] == 'Egypt']
df_alg = df[df['country'] == 'Algeria']
df_mor = df[df['country'] == 'Morocco']
df_tu = df[df['country'] == 'Tunisia']


df_egy.tail()
```

```python
plt.subplots(figsize=(16,8))
plt.plot(df_egy['year'],df_egy['exch_ccy_usd'],label = 'Egypt')
plt.plot(df_tu['year'],df_tu['exch_ccy_usd'], label = 'Tunisia')
plt.plot(df_mor['year'],df_mor['exch_ccy_usd'], label = 'Morocco')
plt.plot(df_alg['year'],df_alg['exch_ccy_usd'], label = 'Algeria')
plt.title('USD Price of North African Currencies from 1860 to 2014')
plt.xlabel('Year')
plt.ylabel('USD Price')
plt.legend()
```
