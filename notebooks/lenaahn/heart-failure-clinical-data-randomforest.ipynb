```python
# Data Prepare
features = ['time', 'ejection_fraction', 'creatinine_phosphokinase', 'age']
hf_target = heart_failure['DEATH_EVENT']
hf_data = heart_failure[features]
hf_data.head()
```

```python
# Data Split
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(hf_data, hf_target,random_state=0)
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_validate
rf = RandomForestClassifier(n_estimators=20, criterion='entropy', n_jobs=-1,
                           random_state=0, oob_score=True)
scores = cross_validate(rf, train_input, train_target,
                       return_train_score=True, n_jobs=-1)
rf.fit(train_input, train_target)
print("Feature Importances")
print(rf.feature_importances_)

print("\n\nTrain Score:")
print(np.mean(scores['train_score']))
print("OOB Score:")
print(np.mean(rf.oob_score_))
print("Test Score:")
print(np.mean(scores['test_score']))
```

```python
# Grid Search
from scipy.stats import uniform, randint
params = {'n_estimators':range(10,100,10),
          'min_impurity_decrease':uniform(0.0001, 0.001),
         'max_depth':randint(20,50),
         'min_samples_split':randint(2,25),
         'min_samples_leaf':randint(1,25)}
from sklearn.model_selection import RandomizedSearchCV
rs = RandomizedSearchCV(RandomForestClassifier(random_state=0), params,
                       n_iter=100, n_jobs=-1, random_state=0)
rs.fit(train_input, train_target)
```

```python
print(rs.best_params_)
```

```python
# Test Score with Best Parameters
print("Test Score:")
print(np.max(rs.cv_results_['mean_test_score']))
```
