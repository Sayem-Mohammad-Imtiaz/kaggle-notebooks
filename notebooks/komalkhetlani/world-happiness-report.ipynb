```python
import matplotlib.pyplot as plt
plt.plot(df1['Happiness Score'],df1['Economy (GDP per Capita)'])
plt.title('Happiness Score vs Economy(GDP per Capita)')
plt.xlabel('Happiness Score')
plt.ylabel('Economy(GDP per Capita)')
plt.show()
```

```python
plt.plot(df1['Happiness Score'],df1['Family'])
plt.title('Happiness Score vs Family')
plt.xlabel('Happiness Score')
plt.ylabel('Family')
plt.show()
```

```python
df1.groupby('Region')['Happiness Score'].mean()
```

Australia and New Zealand  Region has the maximum average Happiness Score while Sub-Saharan Africa has the lowest average Happiness Score.

```python
df1[['Country','Region','Happiness Score']].sort_values(by=['Happiness Score'], ascending=False)
```

Switzerland in Western Europe has the highest happiness score hence the happiest country while Togo in Sub- Saharan Africa has the lowest Happiness Score and hence the least happy country.

```python
top5_2015=df1[['Country','Happiness Rank', 'Happiness Score']].head()
plt.bar(top5_2015['Country'],top5_2015['Happiness Score'])
for index, value in enumerate(top5_2015['Happiness Score']):
    plt.text(index-0.25,value, str(value))
```
