```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings('ignore')
import plotly.graph_objects as go
pd.set_option('display.max_rows',20000, 'display.max_columns',100)
```

```python
mumbai_houseprice = pd.read_csv("../input/housing-prices-in-metropolitan-areas-of-india/Mumbai.csv")
mumbai_houseprice.head()
```

```python
mumbai_houseprice.info()
```

```python
mumbai_houseprice.isnull().sum()
```

```python
mumbai_houseprice['Location'].nunique()
```

## Listed Properties on Makaan.com in Mumbai Region

```python
mumbai_houseprice['Price'] = mumbai_houseprice['Price'].astype(float)
mumbai_houseprice['No. of Bedrooms'] = mumbai_houseprice['No. of Bedrooms'].astype(float)
fig = px.treemap(mumbai_houseprice, path=['Location','Area','Price','No. of Bedrooms'],color='Location')
fig.update_layout(
    title='Listed Properties on Makaan.com in Mumbai Region')
fig.show()
```

## Listing on website from an Mumbai Area

```python
mumbai = mumbai_houseprice.groupby(['Location','Price'])['Area'].sum().reset_index().sort_values('Price',ascending =False)
fig = px.bar(mumbai_houseprice[:700], y='Area', x='Location', color='Price', height=600)
fig.update_layout(
    title='Listing on website from an Mumbai Area ')
fig.show()
```

## Mumbai Region House Price with Location, Area and Bedrooms register on Makaan.com

```python
fig = px.scatter(mumbai_houseprice, x = 'Location',y='Area',color='Price')
fig.update_layout(
    title='Mumbai Region House Price with Location, Area and Bedrooms register on Makaan.com')
fig.show()
```

## Mumbai Region Top 100 House Price with Location and Bedrooms

```python
mumbai= mumbai_houseprice.sort_values('Price',ascending=False).head(100)
fig = px.bar(mumbai, x='Location', y='Price',color='Price', height=500, hover_data=['Price','Area','No. of Bedrooms'])
fig.update_layout(
    title='Mumbai Region Top 100 House Price with Location and Bedrooms')
fig.show()
```

## Bedroomwise House Price Ranges

```python
fig = px.box(mumbai_houseprice, x="No. of Bedrooms", y="Price")
fig.update_layout(
    title='Bedroomwise House Price Ranges in Mumbai Region')
fig.show()
```

## Count of Resale-1/New-0 Houses wrt Vaastu Compliant

```python
mumbai = mumbai_houseprice.groupby(['Location','Resale'])['VaastuCompliant'].count().reset_index(name='VaastuCompliant_Houses')
fig = px.sunburst(mumbai,path=['Location','Resale'],values='VaastuCompliant_Houses')
fig.update_layout(
    title='Count of Resale-1/New-0 Houses wrt Vaastu Compliant')
fig.show()
```
