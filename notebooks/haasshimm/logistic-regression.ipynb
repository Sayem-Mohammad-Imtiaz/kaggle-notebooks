**Reading the csv file**

```python
churn_df=pd.read_csv("../input/churn-modelling/Churn_Modelling.csv")
churn_df
```

**Removing unwanted columns from the dataframe**

```python
churn_df.drop(["RowNumber","CustomerId","Surname"],inplace=True,axis=1)
```

```python
churn_df
```

**Checking for columns that are categorical**

```python
s=(churn_df.dtypes=='object')
s
```

**Finding out the data points of the columns**

```python
print(churn_df.Geography.unique())
print(churn_df.Gender.unique())
```

**One Hot Encoding the columns**

```python
geo=pd.get_dummies(churn_df.Geography)
gen=pd.get_dummies(churn_df.Gender)
```

```python
churn_df=pd.concat([churn_df,geo,gen],axis=1)
churn_df
```

**Very Basic EDA**

```python
churn_df.Balance.plot(kind="hist", figsize=(10,6))
```

```python
churn_df.Balance=np.where(churn_df.Balance==0,0,1)
churn_df.Balance.value_counts()
```

**Splitting into training and testing data**

```python
from sklearn.model_selection import train_test_split
X = churn_df.drop(["Exited","Geography","Gender"], axis=1)
y = churn_df["Exited"]
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42)
```

```python
print(X_train.shape)
X_test.shape
```

**Using Min Max Scaler to scale the data for model**

```python
from sklearn.preprocessing import MinMaxScaler
mmscaler=MinMaxScaler()
mmscaler.fit(X_train)
```

```python
X_train_scaled=mmscaler.transform(X_train)
X_test_scaled=mmscaler.transform(X_test)
```

**Using Logistic Regression to classify**

```python
from sklearn.linear_model import LogisticRegression
reg=LogisticRegression()
reg.fit(X_train,y_train)
```

```python
y_train_pred=reg.predict(X_train)
y_test_pred=reg.predict(X_test)
```

**Finding accuracy score of the model**

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train, y_train_pred))
accuracy_score(y_test,y_test_pred)
```
