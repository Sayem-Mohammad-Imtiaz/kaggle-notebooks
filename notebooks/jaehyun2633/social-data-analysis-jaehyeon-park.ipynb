```python
patient2=patient.dropna( subset=[ 'infected_by' ] ) 
# exclude missing value of 'people who infect another'--> infected by 
```

```python
x=patient2['patient_id']
y=patient2['infected_by']
# declare x= patient_id , infected_by
# compare patient_id & people who infect another 
```

```python
x=x.astype(object)
y=y.astype(object)
# unify datatype
```

```python
pt=[]

for i in x:
    i=str(i)
    pt.append(i)
# make list of patient_id 
```

```python
patient2['patient_ob']=pt

# add column 'patient_ob' = pt 
```

```python
pt=[]


for i in patient2['patient_ob']:
    for j in patient2['infected_by']:
        if i==j:
            pt.append(j)
            
# compare patient_id & people who infect another 
```

```python
pt_list = []

for v in pt:
    if v not in pt_list:
        pt_list.append(v)
# compare patient_id & people who infect another         
```

```python
patient3=patient2[patient2['patient_id'].isin(pt_list)] 
#Extract only if infected person is present and create dataframe.
```

```python
x2=patient3.groupby(["province"]).count()["patient_id"]
# count confirmed_id & groupby 'province' count confirmed person 
```

```python
x2.to_frame() # Series to Dataframe
```

```python
x2=x2.reset_index().rename(columns={"index":"province"}) # reset index 
```

```python
x2=x2.sort_values(by='patient_id',ascending=False) # sort values with confirmed people
```

```python
plt.figure(figsize=(20,20)) 


sns.set(font_scale=3)
ax = sns.barplot(x='patient_id', y='province', data=x2, orient='h', saturation=0.7)
ax.axes.set_title("Region with infected by count", fontsize=40, y=1.01)
ax.set(xlabel='province', ylabel='count');

# show count of infect people by Region(Province)
```
