# Resumo
Nesta análise eu busquei encontrar inconsistências nos dados, para reescrever a tabela omitindo ou removendo informações. A finalidade foi de visualizar possíveis correlações entre a coluna "price" e as demais colunas.

```python
dados=pd.read_csv('../input/craigslist-carstrucks-data/vehicles.csv')
```

# Tratamento dos dados

```python
msno.matrix(dados)
plt.show()
```

```python
dados.head()
```

```python
#removendo colunas "desnecessárias"
dados=dados.drop(columns=['url', 'region_url','image_url', 'VIN', 'county','posting_date', 'description', 'title_status','size'])
#adicionando uma coluna idade
dados['age'] = 2021 - dados['year']
```

```python
dados.head()
```

# Análise de preço x ...


```python
dados.price.describe()
```

```python
#boxplot preços
plt.figure(figsize=(5,10))
sns.boxplot(x=dados.state,y=dados.price)
plt.show()
```

```python
#selecionando apenas preços entre 300 e 70000 dólares
#os valores foram escolhidos por conta do grande número de outliers no boxplot
amostra_precos=dados.loc[(dados.price>=300) & (dados.price<=80000)]
amostra_precos.price.describe()
```

```python
#boxplot preços
plt.figure(figsize=(5,10))
sns.boxplot(y=amostra_precos.price)
plt.title('Boxplot dos preços entre 300usd e 80,000usd')
plt.show()
```

```python
"""#deletar boxplot odometer 
plt.figure(figsize=(5,10))
sns.boxplot(y=amostra_precos.odometer)
plt.ylim(1000,320000)
plt.show()"""


#preco por estado
plt.figure(figsize=(16,8))
sns.boxplot(x=amostra_precos.state, y=amostra_precos.price)
plt.title('Boxplot dos preços por estado entre 300usd  e 80,000usd)')
plt.show()
```

```python
#agrupando preço por idade do carro
k=amostra_precos.groupby('age').agg({'price' : 'mean'})
k.reset_index(inplace=True)
#preco medio por ano
plt.scatter(x=k.age,y=k.price)
plt.xlim(1,41)#ignorando carros muito velhos ou "novos"
plt.title('preço médio dos carros com idades entre 1 e 41 anos')
plt.show()
```

```python
#agrupando odômetro por idade do carro
j=amostra_precos.groupby('age').agg({'odometer' : 'mean'})
j.reset_index(inplace=True)

#preço médio x odometro médio
plt.scatter(x=j.odometer,y=k.price)
plt.xlim(1000,320000) #ignorando valores absurdos
plt.title('preço médio x odômetro médio, ambos agrupados por idade do carro')
plt.show()
```

```python
#condição x preço
ax = sns.barplot(x="condition", y="price", data=amostra_precos)
```

```python
fig = plt.figure(figsize=(10,10))
ax = plt.axes(projection="3d")

tresde= dados.loc[(dados.price>=300)&(dados.price<=80000)&(dados.age>=1)&(dados.age<=41)&(dados.odometer>=1000)&(dados.odometer<=320000)]

z_points = tresde['price']
x_points = tresde['odometer']
y_points = tresde['age']
ax.scatter3D(x_points, y_points, z_points, c=z_points, cmap='hsv');

ax.set_xlabel('odometer')
ax.set_ylabel('age')
ax.set_zlabel('price')
plt.title('odômetro x preço x idade')
plt.show()
```
