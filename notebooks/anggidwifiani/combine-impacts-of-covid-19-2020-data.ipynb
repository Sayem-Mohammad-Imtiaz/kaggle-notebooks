# Combine Data with Multiple Resources

These dataset are about impact of COVID-19 pandemic to lobal economy on 2020. These data take from 2 csv datasets (impact COVID-19 ton 170 countries and unemployment rate on OECD countries). These data respect to total deaths, total cases, Human Development Index (HDI), GDP per Capita, unemployment rate, and country names.

```python
#Importing Packages
import pandas as pd
import numpy as np
```

# Processing Data

```python
#Importing Data
data = pd.read_csv('../input/oecd-unemployment-rate-2020/2020 unemployment.csv')
data.head()
```

```python
#Split Year and Month to Make Clear of Times
time = data['TIME'].str.split('-', expand=True)
data['Year'] = time[0]
data['Month'] = time[1]
data.drop('TIME', axis = 1, inplace = True)
```

```python
#Change column names
data.columns = ['Code', 'Indicator', 'Subject', 'Measure', 'Frequency', 'Value', 'Flag', 'Year', 'Month']
```

```python
#Importing second data
data2=pd.read_csv('../input/impact-of-covid19-pandemic-on-the-global-economy/transformed_data.csv')
data2.head()
```

```python
# Split Year and Month to Make Clear of Times
time = data2['DATE'].str.split('-', expand=True)
data2['Year'] = time[0]
data2['Month'] = time[1]
data2.drop('DATE', axis = 1, inplace = True)
```

```python
# Dropping the data of 2019
data2 = data2[data2.Year != '2019'] 
```

```python
#Change column names
data2.columns = ['Code', 'Country', 'HDI', 'TC', 'TD', 'STI', 'POP', 'GDP', 'Year', 'Month']
```

```python
# Merging both the data together
new_data=pd.merge(data,data2,on=['Code','Month', 'Year'])
```
