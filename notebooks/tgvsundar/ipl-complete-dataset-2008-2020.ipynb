Outputs the number of missing values of the datasets

```python
plt.subplot(221)

df['inning'].value_counts().plot(kind='bar', title='inning', figsize=(16,9))

plt.xticks(rotation=0)

plt.subplot(222)

df['over'].value_counts().plot(kind='bar', title='over')

plt.xticks(rotation=0)

plt.subplot(223)

df['ball'].value_counts().plot(kind='bar', title='ball')

plt.xticks(rotation=0)

plt.subplot(224)

df['batsman'].value_counts().plot(kind='bar', title='batsman')

plt.xticks(rotation=0)

plt.show()
```

**Observation :**  Review of IPL between inning ,over,ball,batsman . Here the subplot(221) means it consists of two Rows and two Columns.figsize returns the height and width of a figure.Title attribute returns the title of the figure.

```python
corr = df.corr()
sns.heatmap(corr, annot=True, square=True)
plt.yticks(rotation=2)
plt.show()
```

annot=True makes  Sure that the values of each cell appear on the chart and yticks sets the y-axis value

```python
df.hist(figsize=(10, 12), bins=50, xlabelsize=8, ylabelsize=8);
```

This plots a histogram with width as 10 and height as 12 bins represents the bars of histogram.bins can be changed as per our wish

```python
sns.relplot(x="extra_runs", y="batsman_runs", hue='batting_team', data=df)
```

**Observation:** The plot is more dense in range 0 to 1 below the batsman runs 0 to 1

```python
sns.factorplot("is_wicket",data=df,kind="count")
```

The is_wicket of 1 have highest count

```python
sns.set_palette("gist_rainbow_r")
sns.jointplot(x="extra_runs",y="total_runs",kind="hex",data=df)
plt.show
```
