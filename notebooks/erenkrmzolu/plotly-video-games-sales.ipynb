```python
# read data
df=pd.read_csv('../input/videogamesales/vgsales.csv')
```

```python
df.info()
```

```python
# Droping nan values
df.dropna(how="any",inplace = True)
df.info()
```

```python
# convert dataframe float to int 
df.Year=df.Year.astype(int)
```

```python
# Line Plot(NA_Sales  and EU_Sales vs rank of top 20 games )

#preparing data
df20=df.head(20)
# import graph objects as 'go'
import plotly.graph_objs as go
# Creating trace 1 
trace1=go.Scatter(x=df20.Rank,
                  y=df20.NA_Sales,
                  mode='lines',
                  name='NA_Sales',
                  marker=dict(color='rgba(16,112,2,0.8)'),
                  text=df20.Rank)
trace2=go.Scatter(x=df20.Rank,
                  y=df20.EU_Sales,
                  mode='lines+markers',
                  name='EU_Sales',
                  marker=dict(color='rgba(80,26,80,0.8)'),
                  text=df20.Rank)
data=[trace1,trace2]
layout=dict(title='NA_Sales  and EU_Sales vs rank of top 20 games ',
           xaxis=dict(title='Rank',ticklen=5,zeroline=False))
fig=dict(data=data,layout=layout)
iplot(fig)
```

```python
# Scatter Plot:NA_Sales and EU_Sales , Other_Sales vs  rank of top 20 games

# import graph objects as 'go'
import plotly.graph_objs as go

# creating trace1
trace1=go.Scatter(x=df20.Rank,
                  y=df20.NA_Sales,
                  mode='markers',
                  name='NA Sales',
                  marker=dict(color='rgba(255,128,255,0.8)'),
                  text=df20.Rank)
# creating trace2
trace2=go.Scatter(x=df20.Rank,
                  y=df20.EU_Sales,
                  mode='markers',
                  name='EU Sales',
                  marker=dict(color='rgba(255,128,2,0.8)'),
                  text=df20.Rank)
# creating trace3
trace3=go.Scatter(x=df20.Rank,
                  y=df20.Other_Sales,
                  mode='markers',
                  name='Other Sales',
                  marker=dict(color='rgba(0,200,255,0.8)'),
                  text=df20.Rank)
data=[trace1,trace2,trace3]
layout=dict(title='NA_Sales and EU_Sales , Other_Sales vs  rank of top 20 games',
            xaxis=dict(title='Rank',ticklen=5,zeroline=False),
            yaxis=dict(title='Sales',ticklen=5,zeroline=False)
)
fig=dict(data=data,layout=layout)
iplot(fig)
```

```python
# Bar Plot :  Global_Sales and NA_Sales of top 5 games 
# prepare data frames
df5=df20.head(5)
# import graph objects as 'go'
import plotly.graph_objs as go
# create trace1
trace1=go.Bar(x=df5.Name,
             y=df5.Global_Sales,
             name='Global Sales',
             marker=dict(color='rgba(255,174,255,0.5)',
                        line=dict(color='rgb(0,0,0)',width=1.5)),
             text=df5.Rank)
# create trace2
trace2=go.Bar(x=df5.Name,
             y=df5.NA_Sales,
             name='NA Sales',
             marker=dict(color='rgba(255,255,128,0.5)',
                        line=dict(color='rgb(0,0,0)',width=1.5)),
             text=df5.Rank)
data=[trace1,trace2]
layout=go.Layout(barmode='group')
fig=go.Figure(data=data,layout=layout)
iplot(fig)
```

```python
#  Second Bar Plot:Global_Sales and NA_Sales of top 5 games 

# prepare data frames
d5=df20.head()
# import graph objects as 'go'
import plotly.graph_objs as go
trace1={
    'x':df5.Name,
    'y':df5.Global_Sales,
    'name':'Global Sales',
    'type':'bar'
};
trace2={
    'x':df5.Name,
    'y':df5.NA_Sales,
    'name':'NA Sales',
    'type':'bar'
};
data=[trace1,trace2]
layout={
    'xaxis':{'title':'Top 5 Games'},
    'barmode':'relative',
    'title':'Global_Sales and NA_Sales of top 5 games '
};
fig=go.Figure(data=data,layout=layout)
iplot(fig)
```

```python
# Bubble Plot games rank(first 20) vs Global sales  with NA_Sales(size) and EU_Sales(color) 

# data preparation 

NA_Sales_size=df20.NA_Sales
EU_Sales_color=df20.EU_Sales

data=[
    {
        'y':df20.Global_Sales,
        'x':df20.Rank,
        'mode':'markers',
        'marker':{
            'color':EU_Sales_color,
            'size':NA_Sales_size,
            'showscale':True
            
        },
        'text':df20.Name
    }
]
iplot(data)
```

```python
# Word Cloud: Most uses games genre in 2016

# datapreparation
df2016=df.Genre[df['Year']==2016]
plt.subplots(figsize=(8,8))
wordcloud=WordCloud(
                     background_color='white',
                     width=512,
                     height=384).generate(' '.join(df2016))


plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('graph.png')
plt.show()
```

```python
# Box plot

# data preparation

trace0=go.Box(y=df20.Global_Sales,
              name='Global sales of top 20 games',
              marker=dict(
              color='rgb(12,12,140)'))
trace1=go.Box(y=df20.EU_Sales,
              name='EU Sales of top 20 games',
              marker=dict(
              color='rgb(12,128,128)'))
trace2=go.Box(y=df20.NA_Sales,
              name='NA Sales of top 20 games',
              marker=dict(
              color='rgb(12,27,28)'))


data=[trace0,trace1,trace2]
iplot(data)
```

```python
# scatter plot:it helps to us to see covarience and relation between more than 2 features

# import figure factory
import plotly.figure_factory as ff

#prepare data
df100=df.loc[:100,['Global_Sales','NA_Sales','EU_Sales']]
df100['index']=np.arange(1,len(df100)+1)

#scatter matrix
fig=ff.create_scatterplotmatrix(df100,diag='box',index='index',colormap='Portland',
                               colormap_type='cat',
                               height=700,
                               width=700)
iplot(fig)
```

```python
# Inset Plot: 2 plots are in one frame
# first line plot
# first line plot
trace1 = go.Scatter(
    x=df20.Rank,
    y=df20.Global_Sales,
    name = "Global Sales",
    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
)
# second line plot
trace2 = go.Scatter(
    x=df20.Rank,
    y=df20.EU_Sales,
    xaxis='x2',
    yaxis='y2',
    name = "income",
    marker = dict(color = 'rgba(160, 112, 20, 0.8)'),
)
data = [trace1, trace2]
layout = go.Layout(
    xaxis2=dict(
        domain=[0.6, 0.95],
        anchor='y2',        
    ),
    yaxis2=dict(
        domain=[0.6, 0.95],
        anchor='x2',
    ),
    title = 'Global Sales and EU Sales vs  Rank of top 20 Games'

)

fig = go.Figure(data=data, layout=layout)
iplot(fig)
    
```

```python
# 3D Scatter Plot

# prepare data
# get top 50 games
df50=df.head(50)

# create trace 1 that is 3d scatter
trace1 = go.Scatter3d(
    x=df50.Rank,
    y=df50.Global_Sales,
    z=df50.EU_Sales,
    mode='markers',
    marker=dict(
        size=10,
        color='rgb(203,0,0)',                # set color to an array/list of desired values      
    )
)
data = [trace1]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0  
    )
    
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
# Multiple Subplots :While comparing more than one features, multiple subplot

# prepare data 
df100=df.head(100)
trace1 = go.Scatter(
    x=df100.Rank,
    y=df100.Global_Sales,
    name = "Global Sales"
)
trace2 = go.Scatter(
    x=df100.Rank,
    y=df100.EU_Sales,
    xaxis='x2',
    yaxis='y2',
    name = "EU Sales"
)
trace3 = go.Scatter(
    x=df100.Rank,
    y=df100.NA_Sales,
    xaxis='x3',
    yaxis='y3',
    name = "NA Sales"
)
trace4 = go.Scatter(
    x=df100.Rank,
    y=df100.Other_Sales,
    xaxis='x4',
    yaxis='y4',
    name = "Other Sales"
)
data = [trace1, trace2, trace3, trace4]
layout = go.Layout(
    xaxis=dict(
        domain=[0, 0.45]
    ),
    yaxis=dict(
        domain=[0, 0.45]
    ),
    xaxis2=dict(
        domain=[0.55, 1]
    ),
    xaxis3=dict(
        domain=[0, 0.45],
        anchor='y3'
    ),
     xaxis4=dict(
        domain=[0.55, 1],
        anchor='y4'
         ),
    yaxis2=dict(
        domain=[0, 0.45],
        anchor='x2'
    ),
    yaxis3=dict(
        domain=[0.55, 1]
    ),
    yaxis4=dict(
        domain=[0.55, 1],
        anchor='x4'
    ),
     title = 'Global Sales, NA Sales , EU Sales and Other Sales  vs World Rank of top 100 games'
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
