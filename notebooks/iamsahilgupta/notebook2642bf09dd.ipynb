```python
import os
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
```

```python
file_path = '../input/heart-attack-analysis-prediction-dataset/heart.csv'
data = pd.read_csv(file_path)
```

```python
features = ['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh', 'exng', 'oldpeak', 'slp', 'caa']

X = data[features]
y = data.output

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)
```

```python
heart_model = DecisionTreeRegressor(max_leaf_nodes=152, random_state=0)
heart_model.fit(train_X, train_y)

predictions = heart_model.predict(val_X)
print(predictions)
print(mean_absolute_error(val_y, predictions))
```
