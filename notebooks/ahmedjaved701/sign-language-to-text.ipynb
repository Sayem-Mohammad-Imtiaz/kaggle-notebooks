```python
test_df = pd.read_csv('/kaggle/input/sign-language-mnist/sign_mnist_test.csv')
```

```python
train_df = pd.read_csv('/kaggle/input/sign-language-mnist/sign_mnist_train.csv')
```

```python
train_df['label'].unique()
```

```python
train_df.shape
```

```python
train_df[train_df['label'] == 1]
```

```python
train_X = train_df.iloc[:,1:].values
train_y = train_df.iloc[:,0].values
test_X = test_df.iloc[:,1:].values
test_y = test_df.iloc[:,0].values
```

```python
train_X_norm = train_X / 255
```

```python
train_X_norm = train_X_norm.reshape(-1, 28, 28)
```

```python
train_X_norm[0]
```

```python
train_y.shape
```

```python
train_y = keras.utils.to_categorical(train_y)
test_y = keras.utils.to_categorical(test_y)
```

```python
train_y.shape
```

```python
model = tf.keras.Sequential(
  [
    tf.keras.layers.Flatten(input_shape=[28,28,1]),
    tf.keras.layers.Dense(512, input_shape=(784,), activation='relu'),
    tf.keras.layers.Dense(25, activation='softmax')
  ])

model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# print model layers
model.summary()
# utility callback that displays training curves
```

```python
train_y.shape
```

```python
train_X.shape
```

```python
model.fit(train_X_norm, train_y, epochs=10)
```

```python
y_pred = (model.predict(test_X) > 0.5).astype("int32")
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test_y,y_pred)
```
