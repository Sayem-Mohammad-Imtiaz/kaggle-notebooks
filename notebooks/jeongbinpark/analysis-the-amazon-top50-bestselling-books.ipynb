## 1.The price change by year

```python
fig =px.box(df, x="Year", y="Price", title="The price's change by year",points="all")
fig.show()
```

### The price almost have range 0-20. The book's price is not change for 10 years

```python
b = pd.DataFrame(df.groupby(["Year"])["Price"].median())

fig = go.Figure()
fig.add_trace(go.Scatter(x=b.index, y=b["Price"],mode="markers+lines",text=b["Price"]))
fig.update_layout(title="The Price change by year - use median")
fig.show()
```

### You can understand this line chart. I use the Price's median. Look. the price's median have the range 8 ~ 13. The market price increase for 10 years. but the book's price is similar for 10 years. Even after 2012, the price is decrease until 2018.

## 2. Compare the Genre by price

```python
fig =px.box(df, x="Year", y="Price", title="The price's change by year + Genre",color="Genre")
fig.show()
```

### Look this box chart. Non Fiction books is more expensive than Fiction books But It's not much of a difference.

```python
a =pd.DataFrame(df.groupby("Genre")["Price"].mean())
a["Price"] = round(a["Price"],2)
b =pd.DataFrame(df.groupby("Genre")["Price"].median())
b["Price"] = round(b["Price"],2)

fig =px.bar(a, x="Price", y=a.index, title="Use Mean",color=['red','blue'],text="Price")
fig.update_traces(textposition='outside')
fig.show()

fig =px.bar(b, x="Price", y=b.index, title = "Use Median",color=['red','blue'],text="Price")
fig.update_traces(textposition='outside')
fig.show()
```

### Also you can check the Non Fiction book is more expensive than Fiction book both mean and median. It have about 3 dollar difference

## 44. User Ratings relationship with Reviews and Price

### I can't find the big relationship. User Ratings is 3.8 and 3.9 that have many Reviews. and 4.1 also have many reviews morethan others

## 6. Change the genre's rate

```python
fig = px.bar(df.sort_values(by="Year"),x="Genre",animation_frame="Year",color="Genre",range_y=(0,40))
fig.update_layout(barmode='stack')
fig.update_xaxes(categoryorder='category ascending')
fig.show()
```

```python
a = df.melt(["Year","Genre"])
a["percent"] = 1/250
px.bar(a, x="Year",y="percent", color="Genre",color_discrete_sequence=["red","blue"])
```

### you can check that only 2014 is Fiction books more than Non Fiction. other year is Non Fiction book is more than Fiction
