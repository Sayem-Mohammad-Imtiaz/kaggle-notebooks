```python
data = pd.DataFrame(df.groupby('type')["title"].count())
data.reset_index(inplace=True)
data.columns=["type","value"]
fig = px.pie(data,values="value",names="type",title="Total ratio of program type")
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

### Totally, Movie is more than Tv Show. Movie is 69.1% and Tv Show is 30.9%.

```python
df['tmp']=1

data = df.pivot_table(index=["type","release_year"],values="tmp",aggfunc="sum",fill_value=0)

data = data.reset_index()

fig = px.bar(data[data['release_year']>2000],x="release_year",y="tmp",color="type",text="tmp")
fig.update_layout(barmode='group')
fig.update_traces(textposition='outside')
fig.show()
```

```python
df["date_added"] = df['date_added'].fillna(df["release_year"])

def make_added(date_added):
    if type(date_added)==int:
        return date_added
    else:
        return date_added.split(",")[1][1:]

df["date_added_1"] = df["date_added"].map(make_added)

df["date_added_1"]= df["date_added_1"].astype(int)

data = df.groupby("date_added_1")[['show_id']].count()

data.reset_index(inplace=True)

data.columns=["year","count"]

fig = px.bar(data,x="year",y="count",text="count",color="count",)
fig.update_traces(textposition='outside')
fig.show()
```

### After 2016, Netflix is shown anormous growth.

```python
country = df[~df["country"].isnull()]['country'].to_list()

country = str(country)

country = re.sub("[\'\]\[]","",country)
country = re.sub(" {2,}", " ", country)

country = pd.Series(country.split(",")).value_counts()

country = pd.DataFrame(country)
country.reset_index(inplace=True)

country.columns=["country","count"]

fig = px.bar(country.head(20), x="country", y="count",title="Top 20 Netflix country",text="count",color="country")
fig.update_traces(textposition="outside")
fig.show()
```

###  Netflix record best Number of Movies and Tv Shows on United States. and India, United Kingdom followed. Also in other countries, Netflix make me feel that it is a global company

```python
def no_space(sen):
    return sen[1:]

country["country"]=country["country"].apply(no_space)

from iso3166 import countries

countries = pd.DataFrame(countries)

countries.drop(["alpha2","numeric","apolitical_name"],axis=1,inplace=True)

num = countries.to_numpy()

def make_alph3(cou):
    for i in range(len(num)):
        if cou == num[i][0]:
            return num[i][1]
    
    return 'no'

country["country_iso3"] = country["country"].apply(make_alph3)

country.drop([64,99],inplace=True)

country.loc[0,"country_iso3"] = "USA"
country.loc[2,"country_iso3"] = "GBR"
country.loc[7,"country_iso3"] = "KOR"
country.loc[18,"country_iso3"] = "TWN"
country.loc[35,"country_iso3"] = "RUS"
country.loc[41,"country_iso3"] = "CZE"
country.loc[59,"country_iso3"] = "VNM"
country.loc[61,"country_iso3"] = "DEU"
country.loc[65,"country_iso3"] = "IRN"
country.loc[68,"country_iso3"] = "RUS"
country.loc[74,"country_iso3"] = "VEN"
country.loc[89,"country_iso3"] = "DEU"
country.loc[106,"country_iso3"] = "VAT"
country.loc[113,"country_iso3"] = "SYR"

fig = px.choropleth(country, locations="country_iso3",
                    color="count",
                    hover_name="country",
                    range_color=(0, 10),
                    color_continuous_scale="Blues",)
fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

# Movie's Run Time
</h1>

```python
movie = df[df["type"] =='Movie']

def make_min(minit):
    return minit.split(" ")[0]

movie["duration"] = movie["duration"].apply(make_min)

movie["duration"] = movie["duration"].astype(int)

movie["duration"].describe()

px.histogram(movie["duration"], range_x = (0,200),title="Movie's Run Time")
```

### Almost movie's runtime is 80min ~ 120min, and Average is 99.30 min.
