```python
healthupdatae=health.iloc[:,2:]
healthupdatae
```

```python
healthupdatae['work_type'].unique()
```

```python
worktype= {'Private': 1,'Self-employed': 2,'Govt_job':3,'children':4,'Never_worked':5}
healthupdatae.work_type = [worktype[item] for item in healthupdatae.work_type] 
```

```python
healthupdatae
```

```python
healthupdatae['Residence_type'].unique()
```

```python
area={'Urban':1,'Rural':2}
healthupdatae['Residence_type'] = [area[item] for item in healthupdatae['Residence_type']] 
```

```python
healthupdatae
```

```python
healthupdatae['ever_married'].unique()
```

```python
healthupdatae['ever_married'].replace(('Yes', 'No'), (1, 0), inplace=True)
```

```python
healthupdatae
```

```python
healthupdatae['smoking_status'].unique()
```

```python
smoketype= {'formerly smoked': 1,'never smoked': 2,'smokes':3,'Unknown':4}
healthupdatae.smoking_status = [smoketype[item] for item in healthupdatae.smoking_status] 
```

```python
healthupdatae['smoking_status'].unique()
```

```python
healthupdatae
```

```python
feature_cols =['age','hypertension','heart_disease','ever_married','work_type','Residence_type','avg_glucose_level','bmi','smoking_status']
 # Features
feature_cols
X = healthupdatae[feature_cols] # Features
y = healthupdatae.stroke
```

```python
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
print("Accuracy  for Normal Criterion :",metrics.accuracy_score(y_test, y_pred))
```

```python
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)
clf = clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
print("Accuracy for Criterion Entropy:",metrics.accuracy_score(y_test, y_pred))
```
