```python
import pandas as pd
import numpy as np
```

```python
HR_df = pd.read_csv("../input/hr-analytics/HR_comma_sep.csv")
```

```python
HR_df.head()
```

```python
HR_df.info()
```

```python
HR_df.drop(['Work_accident', 'number_project', 'promotion_last_5years'], axis='columns', inplace=True)
```

```python
HR_df
```

```python
# salary dummies
HR_df_sallary_dummies = pd.get_dummies(HR_df['salary'])
```

```python
# department dummies
HR_df_dummies_department = pd.get_dummies(HR_df['Department'])
```

```python
new_HR_df = pd.concat([HR_df,HR_df_sallary_dummies,HR_df_dummies_department], axis='columns')
```

```python
new_HR_df
```

```python
new_HR_df.drop(['Department', 'salary', 'medium', 'technical'], axis = 'columns', inplace=True)
```

```python
new_HR_df
```

```python
# separate independent variable
X = new_HR_df.drop(['left'], axis='columns')
```

```python
# separate dependent variable
y = new_HR_df['left']
```

```python
# Model training

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35,random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='liblinear', random_state=0)
```

```python
model.fit(X_train, y_train)
```

```python
# score of the model
model.score(X_test,y_test)
```
