```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
        
df = pd.read_csv('/kaggle/input/world-happiness/2019.csv')
```

```python
# Check for any missing data
df.info()
```

```python
# Rename the columns for easy access of data
df.rename( columns = {
    'Overall rank' : 'rank',
    'Country or region' : 'country',
    'Score' : 'score',
    'GDP per capita' : 'gdp',
    'Social support' : 'support',
    'Healthy life expectancy' : 'expectancy',
    'Freedom to make life choices' : 'freedom',
    'Generosity' : 'generosity',
    'Perceptions of corruption' : 'corruption'
}, inplace = True)
```

```python
# A function to plot the points to find their coeeficient
def plot(x, y, label1, label2):
    
    fig, ax = plt.subplots()
    ax.scatter(x, y)
    
    title = f'Correlation between {label1} and {label2}'

    ax.set(
        xlabel=label1,
        ylabel=label2,
        title=title
    )
    
    m, b = np.polyfit(x, y, 1)
    
    coeff = np.corrcoef(x,y)[0][1]
    
    print("Correlation = {0:.2f}".format(coeff))
    print()
    plt.plot(x, m*x + b)
    plt.show()
```

<h2>Are rich countries are more happy</h2>
<p>From the plot below we can see that there is a strong correlation between wealth and happiness</p>

```python
# Are rich countries are more happy
plot(df['gdp'], df['score'], 'gdp', 'score')
```

<h2>Does family support make people happy</h2>
<p>It is a not a suprise to find out that family support makes people happy. This can be generalized by stating that support in general contributes considerable to happiness index of the country.</p>

```python
# Does Family support make people happy
plot(df['support'], df['score'], 'support', 'score')
```

<h2> Does life expectancy makes a person happy</h2>
<p>By life expectancy we mean health of an individual. As we cannot predict how long we will live, it doesnt make sense to state that life expectancy makes people happy. What we can conclude from the observation is that as life expectancy and health is related we can state that <i>health makes a person happy</i>.</p>

```python
# Does life expectancy affect happiness
plot(df['expectancy'], df['score'], 'expectancy', 'score')
```

<h2>Does freedom affect happiness</h2>
<p>We can clearly see that there is a correlation beween freedom and happiness of a country.</p>

```python
# Does freedom affect happiness
plot(df['freedom'], df['score'], 'freedom', 'score')
```

<h2>Generosity does not affect happiness !!</h2>
<p>It is suprising to see that generosity has very little effect on happiness. It may not be the reality as our data size is extremely small to make such a bold claim. Additionally generosity is entirely an subjective term and is very difficult to quantify.</p>

```python
# Does generosity affect happiness
plot(df['generosity'], df['score'], 'generosity', 'score')
```

<h2>Corruption and happiness</h2>
<p>There is a slight correlation between corruption and happiness which is not really suprising</p>

```python
# Does corruption affect happiness
plot(df['corruption'], df['score'], 'corruption', 'score')
```

<h2>Correlation between various parameters</h2>
<p>This is a pairplot that shows the correlation between various parameters</p>

```python
sns.pairplot(df , corner=True, kind='reg', plot_kws={'line_kws':{'color':'green'}})
```
