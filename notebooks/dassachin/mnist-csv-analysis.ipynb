```python
x=train_data.iloc[:,1:]
y=train_data.iloc[:,:1]
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_valid,y_train,y_valid=train_test_split(x,y,test_size=0.2,random_state=25)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import  accuracy_score
model1 = DecisionTreeClassifier()
model1.fit(x_train,y_train)
y_pred1=model1.predict(x_valid)
print(accuracy_score(y_valid,y_pred1)*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
data = KNeighborsClassifier(n_neighbors=5)
data.fit(x_train,y_train)
prediction = data.predict(x_valid)
print(accuracy_score(y_valid,prediction)*100)
```

```python
from xgboost import XGBClassifier
# specify parameters via map
model=XGBClassifier( learning_rate =0.01,
 n_estimators=50,
 max_depth=5)
model.fit(x_train,y_train)
pred1 = model.predict(x_valid)
print(accuracy_score(y_valid,pred1)*100)
```
