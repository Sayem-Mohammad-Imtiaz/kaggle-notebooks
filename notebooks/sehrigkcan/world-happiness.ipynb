```python
_2015=pd.read_csv("/kaggle/input/world-happiness/2015.csv")
_2015.head()
_2015.drop(columns=['Happiness Rank','Standard Error','Economy (GDP per Capita)','Family','Dystopia Residual'],inplace=True) # gereksiz columns sildik
```

```python
_2015.rename(columns={"Happiness Score": "Score", "Health (Life Expectancy)": "Health","Trust (Government Corruption)":"Corruption"},inplace=True)  # columns isimleri degistirdik
_2015['Year'] = list(len(_2015)*[2015])   # Year sutunu ekledik 
_2015
```

```python
_2016=pd.read_csv("/kaggle/input/world-happiness/2016.csv")
_2016.columns
_2016.drop(columns=['Happiness Rank','Lower Confidence Interval','Upper Confidence Interval','Economy (GDP per Capita)','Family','Dystopia Residual'],inplace=True)
_2016['Year'] = list(len(_2016)*[2016])   # Year sutunu ekledik 
_2016.rename(columns={"Happiness Score": "Score", "Health (Life Expectancy)": "Health","Trust (Government Corruption)":"Corruption"},inplace=True)  # columns isimleri degistirdik
_2016.columns
set(_2016.Region)
```

```python
_2017=pd.read_csv("/kaggle/input/world-happiness/2017.csv")
_2017.drop(columns=['Happiness.Rank','Whisker.high','Whisker.high','Whisker.low','Economy..GDP.per.Capita.','Family','Dystopia.Residual'],inplace=True)
_2017.rename(columns={"Happiness.Score": "Score", "Health..Life.Expectancy.": "Health","Trust..Government.Corruption.":"Corruption"},inplace=True)  # columns isimleri degistirdik
_2017['Year'] = list(len(_2017)*[2017]) 


a=list(_2015["Country"])                             # ulke listesi olusturduk 2015teki ulkeleri kullanarak
b=list(_2015["Region"])                               # bolge listesi olusturduk 2015teki bolgeleri kullanarak
c= {k:v for k,v in zip(a,b)}                          # ulke ve bolgeleri bir sozlukte birlestirdik
_2017['Region'] = _2017['Country'].replace(c)         # Region satirini c sozlugune gore ekledik

_2017= _2017[['Country', 'Region', 'Score', 'Health', 'Freedom', 'Corruption','Generosity', 'Year']]       

_2017["Region"].replace({"Taiwan Province of China":"Asia","Belize":"Central America","Hong Kong S.A.R., China":"Asia Pacific","Somalia":"Eastern Africa","Namibia":"Southern Africa","South Sudan":"Northern Africa"},inplace=True) 
_2017
```

```python
_2018=pd.read_csv("/kaggle/input/world-happiness/2018.csv")
_2018.drop(columns=['Overall rank','GDP per capita','Social support'],inplace=True)
_2018['Region'] = _2018['Country or region'].replace(c)
_2018.rename(columns={"Country or region": "Country", "Healthy life expectancy": "Health","Perceptions of corruption":"Corruption", "Freedom to make life choices":"Freedom"},inplace=True)  # columns isimleri degistirdik
_2018= _2018[['Country', 'Region', 'Score', 'Health', 'Freedom', 'Corruption','Generosity']]  
_2018['Year'] = list(len(_2018)*[2018]) 

for x in _2018['Region']:
    for y in _2018['Country']:
        if x==y:
            print(x)
            
_2018["Region"].replace({"Trinidad & Tobago":"Latin America and the Caribbean","Belize":"Central America","Northern Cyprus":"European Union","Somalia":"Eastern Africa","Namibia":"Southern Africa","South Sudan":"Northern Africa"},inplace=True)
_2018.tail(50)
```

```python
_2019=pd.read_csv("/kaggle/input/world-happiness/2019.csv")
_2019.drop(columns=['Overall rank','GDP per capita','Social support'],inplace=True)
_2019['Region'] = _2019['Country or region'].replace(c)
_2019.rename(columns={"Country or region": "Country", "Healthy life expectancy": "Health","Perceptions of corruption":"Corruption", "Freedom to make life choices":"Freedom"},inplace=True)  # columns isimleri degistirdik
_2019= _2019[['Country', 'Region', 'Score', 'Health', 'Freedom', 'Corruption','Generosity']]  
_2019['Year'] = list(len(_2019)*[2019]) 
_2019["Region"].replace({"Trinidad & Tobago":"Latin America and the Caribbean","North Macedonia": "Europe","Northern Cyprus":"European Union","Somalia":"Eastern Africa","Namibia":"Southern Africa","South Sudan":"Northern Africa","Gambia":"Western Africa"},inplace=True)

# replace metodu ile ulke ve bolgeleri ayni olanlari degistirdigimizin kontrolu 
#for x in _2019['Region']:
#    for y in _2019['Country']:
#        if x==y:
#            print(x)

_2019.tail(50)
```
