<div class="alert alert-success" style="max-width:500px;">
    <h1>Splitting the Data<br></h1><b> using simple mask technique , you can also use built in train_test_split function</b>
</div>

```python
mask=np.random.rand(len(df))<0.8
train=df[mask]
test=df[~mask]
```

```python
x=np.asanyarray(train[['fixed acidity','volatile acidity','citric acid','residual sugar','density','pH','alcohol']])
y=np.asanyarray(train['quality'])
```

<div class="" style="max-width:500px;">
    <h1>Train the Model</h1>
</div>

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
model.fit(x,y)
print("Coefficients :" , model.coef_)
print("Intercept" , model.intercept_)
```

```python
test_x=np.asanyarray(train[['fixed acidity','volatile acidity','citric acid','residual sugar','density','pH','alcohol']])
test_y=np.asanyarray(train['quality'])
```

```python
pred=model.predict(train[['fixed acidity','volatile acidity','citric acid','residual sugar','density','pH','alcohol']])
```

<div class="alert alert-danger" style="max-width:500px;">
    <h1>Check Residual sum of Squares Error</h1>
</div>

```python
print('Residual sum of square is %.2f'%np.mean(test_y-pred)**2)
```
