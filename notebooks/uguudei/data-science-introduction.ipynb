### Applying function

```python
mean_charge = data['charges'].mean()
mean_charge
```

```python
def charge_difference(value):
    value_diff = value - mean_charge
    return value_diff

data['charges'].apply(charge_difference)
```

```python
# labmda: nameless function
data['charges'].apply(lambda x: x - mean_charge)
```

```python
data['charge_diff'] = data['charges'].apply(lambda x: x - mean_charge)
data
```

```python
data['mean_charge'] = data['charges'].mean()
display(data)
data['charge_diff'] = data['charges'] - data['mean_charge']
display(data)
```

```python
data['sex_smoker'] = data['sex'] + '_' + data['smoker']
data
```

```python
# Dropping previously created columns
data = data.drop(['charge_diff', 'mean_charge', 'sex_smoker'], axis=1)
data
```

### Group by

```python
# group by on single column
data.groupby('region')['charges'].sum().round()
```

```python
# group by with multiple aggregations
data_grouped = data.groupby(['region']).charges.agg([len, 'mean', min, max])
data_grouped
```

```python
# group by multiple columns with multiple aggregations
data.groupby(['region', 'sex']).charges.agg([len, 'mean', min, max])
```

```python
# Group by with multiple aggregations
import numpy as np
data.groupby('region').agg(
    {'charges':['min', 'mean', 'max'], 
     'bmi':'mean', 
     'age':'mean'
    })
```
