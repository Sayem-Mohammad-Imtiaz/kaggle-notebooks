# Recommender System**********

```python
# lets make a pivot table in order to make rows are users and columns are music.

pivot_table = df.pivot_table(index = ["id"],columns = ["composer"])
pivot_table.head(5)
```

```python
pivot_table.shape
```

Now we will make a function which will recommend the music with their correlation score.Note that higher the correlation more the movie related to each othe

```python
def recommended_music(music):
    music_listen = pivot_table[music]
    similarity_with_other_music = pivot_table.corrwith(music_listen)
    similarity_with_other_music = similarity_with_other_music.sort_values(ascending=False)
    return similarity_with_other_music.head()
```
