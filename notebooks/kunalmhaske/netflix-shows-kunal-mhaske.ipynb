# Shows/Movies Percentage which people are watching******

```python
labels = 'Jay bhavani-JAy Shivaji', 'Euma', 'Empire', 'Master', 'YJHD', 'Ev', 'suchi'
sizes = [200, 50, 90, 100, 60, 80, 140]
explode = (0, 0, 0, 0, 0, 0, 0.1) 

fig1, ax1 = plt.subplots(figsize=(15,10))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal') 
plt.show()
```

```python
plt.figure(figsize=(10,10))
id1 = np.arange(1,10)
salary = np.arange(20,110,10)
plt.bar(id1,salary,color ='#42B300')
plt.xlabel('show_id')
plt.ylabel('type')
plt.show()
```
