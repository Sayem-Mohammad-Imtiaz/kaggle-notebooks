# Global Map for Covid-19 Vaccinations

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv', header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
```

Below is the world map shows percentage of people get vaccinated by each countries. we are now can see people in most countries in Central Africa and Meddle East got less than 5% of its population by now and the daily vaccination informaiton is also not available.    

```python
fig = px.choropleth(df,locations='country', locationmode='country names', color='people_vaccinated_per_hundred',hover_name='country',animation_frame='date', color_continuous_scale='Greens', range_color=(0.0,60.0))
fig.update_layout(title_text="Time frame of the country's population got vaccination in percentage", title_x=0.5)
fig.show()

plotly.offline.plot(fig,filename='vaccinations.html')
```

# Earthquake is waking up in Iceland
News source: In Iceland, 18,000 Earthquakes Over Days Signal Possible Eruption on the Horizon, the New York Tims, March 4, 2021 https://www.nytimes.com/2021/03/04/world/europe/earthquakes-eruption-iceland.html

According to the news, volcanoes in southwestern Iceland have been quiet for 800 years. Just recently from february 2021,18,000 earthquakes have been reported and most of them happend near residents in the capital, Reykjavik. The largest one was happen in February 24 with a 5.7-magnitude quake and thousands of others have since followed. 

```python
df = pd.read_csv('../input/2021-iceland-recent-earthquakes/query (1).csv',header=0)
df.index = pd.to_datetime(df['time'])
df['time']=df.index.strftime('%y-%m-%d-%H:00:00')
df = df.sort_index()
fig = px.scatter_geo(df,lat='latitude',lon='longitude',color= 'mag',animation_frame='time', color_continuous_scale='Bluered',range_color=(1.0,6.0))
fig.show()
plotly.offline.plot(fig,filename='21 Feb-Mar Earthquakes.html')
```
