```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

selected_country_1='Cambodia'
selected_country_2='Cambodia'

df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1 = df1[df1['country']==selected_country_1]
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2 = df2[df2['country']==selected_country_2]
df1.index = pd.to_datetime(df1['date'])
df2.index = pd.to_datetime(df2['date'])
```

**The Expos√© to the possible cause of Feb 20 incident**

News sources: 
* https://www.khmertimeskh.com/50823790/escape-from-sokha-hotel-the-expose-to-the-possible-cause-of-feb-20-incident/
* https://www.b2b-cambodia.com/articles/february-20-2021-community-incident-phnom-penh/


Cambodia has struggled for a so-called mass infection of COVID-19 on Februay 20 2021, which believed that it was a huge community incident of all times. According to the news, the chaos incident was caused by the escape from the quarantine accommodations by a group of foreigners after arrival, and later they were found positive of COVID-19. After that, the capital city was lockdown and divided into red (very high risk), yellow (medium risk), and orange (average risk) zone. At the same time, the government also started to provide vaccination campaigns across the country especially in the capital city. 

```python
fig,ax = plt.subplots(figsize=(10,5))
ax.plot(df1.index,df1['daily_new_cases'],label='Cases')
ax1 = ax.twinx()
ax1.plot(df2.index,df2['daily_vaccinations'],'r', label='Vaccinations')
```

```python
from plotly.offline import iplot
import plotly.graph_objs as go
daily_new_cases = go.Scatter(x=df1.index, y=df1['daily_new_cases'], yaxis='y1', name='Daily confirmed cases')
daily_vaccinations = go.Scatter(x=df2.index, y=df2['daily_vaccinations'], yaxis='y2', name='Daily vaccinations')
layout_obj = go.Layout(title='Cambodia: Vaccination and Infection from Feb. 20', xaxis=dict(title='Date'), yaxis=dict(title='Daily cases'), yaxis2=dict(title='Vaccinations', side='right', overlaying='y'))
fig = go.Figure (data=[daily_new_cases,daily_vaccinations], layout=layout_obj)
iplot(fig)
fig.write_html('outcome.html')
```

**Global Top 10 Countries Ranking of Daily New Cases**

```python
df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1 = df1.groupby('country').sum()['daily_new_cases'].sort_values(ascending=False) [0:10]
print(df1.to_string())
```

**Global Top 10 Countries Ranking of Vaccinations**

```python
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2 = df2.groupby('country').max()['people_vaccinated_per_hundred'].sort_values(ascending=False) [0:10]
print(df2.to_string())
```
