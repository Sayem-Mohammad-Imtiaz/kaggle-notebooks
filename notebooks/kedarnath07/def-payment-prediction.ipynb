```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

# 4.Data Preprocessing

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
std_x = std.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.21,random_state = 41)

print("Training data:{}".format(x_train.shape))
print("Test data:{}".format(x_test.shape))
```

# 4.Built Model

```python
from sklearn.linear_model import SGDClassifier

clf = SGDClassifier()
```

```python
from sklearn.model_selection import RandomizedSearchCV
#Randomized Search CV


loss = ['hinge', 'log', 'modified_huber','squared_hinge', 'perceptron','squared_loss']

penalty = ['l2', 'l1']

alpha = [0.001,0.0001,0.00025,0.0000025,2,5,10,20]

max_iter = [10000,10000,20000,55000,40000]

#learning_rate = ['constant','optimal','invscaling','adaptive']
```

```python
# Create the random grid
random_grid = {'loss': loss,
               'penalty': penalty,
               'alpha': alpha,
               'max_iter': max_iter}

print(random_grid)
```

```python

# Random search of parameters, using 3 fold cross validation, 
# search across 50 different combinations
rf_random = RandomizedSearchCV(estimator = clf, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = 1)
rf_random.fit(x_train,y_train)
```

```python
print(rf_random.best_params_)
```

```python
clf = SGDClassifier(penalty ='l1', max_iter= 40000, loss = 'log', alpha = 0.001)

clf.fit(x_train,y_train)

score =  clf.score(x_test,y_test)
print(score)
```
