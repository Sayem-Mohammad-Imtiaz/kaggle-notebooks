# Simple List Series with Pandas Lib

```python
List = [1, 2 ,3 ,4 ,5 ,6]
series = pd.Series(List)
```

```python
print(series)
```

# Return 4th Value Of List

```python
print(series[4])
```

# Dfine Special index

```python
List_01 = [9 ,7 ,8]
data_01 = pd.Series(List_01 ,index = ["AI" , "ML" , "DL"])
```

```python
print(data_01)
```

```python
print(data_01['ML'])
```
