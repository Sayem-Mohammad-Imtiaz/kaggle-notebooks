# **1.Background:** 
* As of March 26, UTC, the cumulative number of people infected with COVID-19 has been no less than 6,000,000 and the number of deaths has exceeded 30,000. 
* At the same time, COVID-19 has spread in more than 200 countries and regions. There is no doubt that this disaster has become a disaster for all mankind. No country, no one can survive alone.  

1.1 According to the current confirmed diagnosis data, patients are mostly distributed in countries with higher economic levels and stronger medical testing capabilities. The population of developing countries accounts for 80% of the world's population, but the number of infections in developing countries is significantly lower in developing countries than in developing countries. Limited by the three factors of medical conditions, education, and economic development, we can think that infection data in developing countries are generally greatly underestimated. If the epidemic is concentrated in developing countries, the weak medical system in developing countries will face the risk of rapid collapse, and the number of deaths will rise sharply. At the same time, the fragile economic and social order in developing countries will be quickly defeated. Numerous secondary disasters such as war and hunger will lead to humanitarian disasters worldwide (and even our inability to rescue). Based on this situation, we want to solve two problems: What is the spread of COVID-19 in different countries around the world? Which countries have their data underestimated because of testing capabilities or other factors?

* First of all, it is certain that developing countries have very few testing capabilities and medical resources. At the same time, factors such as economic backwardness and inconvenient transportation make it difficult for these people to get testing opportunities. For this reason, testing data will be far lower than the true value .
* But what is the "true value" of the number of people infected in developing countries? On this issue, I think we need to consider the impact of two different factors in developing countries on the conclusion.
* Considering positive factors, economic backwardness and inconvenient transportation in developing countries will delay the spread of COVID-19. A lower degree of urbanization means a lower population density, and a lower economic level means a lower flow of people with foreign countries. These are factors that slow the spread of COVID-19 in developing countries.
Considering negative factors, the level of education in developing countries is generally low, and information transmission channels are backward, which means that many people cannot understand how to avoid getting infected with the virus. At the same time, a lower economic level means that the average physical fitness of the population, that is, immunity will be lower, and COVID-19 is more likely to be infected.
