# Flexible arguments

```python
# *args
def f(*args):
    for i in args:
        print(i)
f(1)
f("")
f(1,2,3,7)
```

```python
f("asdasdasdas",4545)
```

```python
# **kwargs
def f(**kwargs):
    """print key and value of dictionary"""
    for key,value in kwargs.items():
        print("key:",key,"value:",value)
f(country = 'spain', capital = 'madrid', population = 123456)
```
