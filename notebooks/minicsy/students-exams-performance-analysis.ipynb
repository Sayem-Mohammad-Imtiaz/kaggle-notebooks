```python
# Bar plot by gender
plt.figure(figsize=(6,6))
sns.countplot(x='gender', data=df)

ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2, p.get_height(), '%d' % int(p.get_height()),
    fontsize=16, color='red', ha='center', va='bottom')

plt.show()
```

```python
# Bar plot by race/ethnicity
plt.figure(figsize=(12,6))
df['race/ethnicity'].value_counts().plot.bar()

ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2, p.get_height(), '%d' % int(p.get_height()),
    fontsize=16, color='red', ha='center', va='bottom')

plt.show()
```

```python
# Bar plot by race/ethnicity
plt.figure(figsize=(12,6))

cmap = sns.color_palette('Set2')
sns.countplot(x='race/ethnicity',data=df, palette=cmap);
plt.xticks(rotation=45);

ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2, p.get_height(), '%d' % int(p.get_height()),
    fontsize=16, color='red', ha='center', va='bottom')

plt.show()
```

```python
# Bar plot by race/ethnicity
plt.figure(figsize=(12,6))

cmap = sns.color_palette('Set2')
sns.countplot(x='parental level of education',data=df, palette=cmap);
plt.xticks(rotation=45);

ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2, p.get_height(), '%d' % int(p.get_height()),
    fontsize=16, color='red', ha='center', va='bottom')

plt.show()
```

```python
# Count plot by race/ethnicity
plt.figure(figsize=(12,6))
sns.countplot(df['race/ethnicity'])

ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2, p.get_height(), '%d' % int(p.get_height()),
    fontsize=16, color='red', ha='center', va='bottom')

plt.title('Number of Students by Race/Ethnicity')
plt.ylabel('Number of Students')
plt.xlabel('Race/Ethnicity')
plt.xticks(rotation = 90)
    
plt.show()
```

```python
# Count plot by parental level of education
plt.figure(figsize=(12,6))
sns.countplot(df['parental level of education'])
ax = plt.gca()
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2, p.get_height(), '%d' % int(p.get_height()),
    fontsize=16, color='red', ha='center', va='bottom')

plt.title('Number of Students by Parental Level of Education')
plt.ylabel('Number of Students')
plt.xlabel('parental level of education')
plt.xticks(rotation = 90)
    
plt.show()
```
