```python
df.Open.plot(kind='line', color='g',label='Open',linewidth=1, alpha=0.4,grid=True,linestyle=':')
df.Close.plot(color='r',label='Close',linewidth=1, alpha=0.4,grid=True,linestyle='-.')
plt.legend(loc='upper right')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Line Plot')
```

```python
df.plot(kind='scatter', x='Open', y='Close',alpha = 0.5,color = 'red')
plt.xlabel('Open')             
plt.ylabel('Close')
plt.title('Open Close Scatter Plot') 
```
