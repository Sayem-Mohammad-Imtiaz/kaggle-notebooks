##### Индивидуальный проект, часть 1
*Выполнено Волковой Ульяной, группа суббота*

В проекте использован массив данных, собранный в Университете Висконсина в начале 1990-х годов, характеризующий основные потенциальные индикаторы рака груди, в особенности толщину скопления клеток (доброкачественные клетки располагаются монослойно, в то время как злокачественные формируют мультислои), оценки размеров отдельных клеток эпителия, единичные "голые" ядра (типичные для клеток доброкачественных тканей), динамику митоза и прочие свойства.
Актуальность выбора задачи состоит главным образом в том, что с раком молочной железы сталкивается все больше женщин, и в свете стремительного распространения этого заболевания необходимо более оптимальные и точные методы его детекции на ранних стадиях.

Информация о параметрах:

1) Идентификационный номер образца
2) Диагноз (M - злокачественное, B - доброкачественное новообразование)
3) - 32)

10 оцениваемых параметров для каждого клеточного ядра, полученных с помощью программы Xcyt:

1) радиус
2) текстура (в рамках стандартного отклонения по шкале серого)
3) периметр
4) площадь
5) гладкость
6) компактность ($$dfrac{периметр^2}{площадь - 1.0}$$)
7) вогнутость (выраженность вогнутых участков по контуру ядра)
8) количество вогнутых участков
9) симметрия
10) фрактальная размерность (аппроксимация длины кривой по Мандельброту

Медиана, стандартная ошибка и "худшее" или наибольшее значения (медиана трех наибольших значений) каждого параметра были вычислены для каждого изображения и в результате получилось 30 характеристик, а все значения округлены до 4 знаков после запятой. Пропущенных значений нет. Классовое распределение: 357 доброкачественных и 212 злокачественных.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import pylab 
import scipy.stats as stats
```

```python
df = pd.read_csv("../input/data.csv",header = 0)
df.head()
```

```python
df.iloc[:,1:].describe()
```

```python
normalised = (df["radius_mean"] - df["radius_mean"].mean())/df["radius_mean"].std()
d = {"radius_mean": normalised, "diagnosis": df["diagnosis"]}
norm_df = pd.DataFrame(data = d)
norm_df.head()
```

```python
stats.probplot(norm_df["radius_mean"], dist="norm", plot=pylab)
pylab.show()
```

Распределение не нормальное, как очевидно из графика выше.

```python
print(stats.shapiro(df["radius_mean"]))
print(stats.shapiro(normalised))
```

Согласно критерию Шапиро-Уилка данные распределены не нормально.

```python
stats.wilcoxon(df[df["diagnosis"] == 'M']["radius_mean"].sample(n=200) \
               , df[df["diagnosis"] == 'B']["radius_mean"].sample(n=200))
```

Согласно критерию Уилкоксона радиусы доброкачественных и злокачественных клеток различаются значимо.
