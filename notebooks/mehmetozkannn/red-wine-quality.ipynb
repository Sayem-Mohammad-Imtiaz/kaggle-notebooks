```python
df = pd.read_csv("/kaggle/input/red-wine-quality-cortez-et-al-2009/winequality-red.csv")
df.head()
```

```python
myColumn = df["total sulfur dioxide"]
myColumn.shape
```

# Explanation


```python
meanOfNumbers = myColumn.mean()
meanOfNumbers
```

```python
medianOfnumbers = myColumn.median()
medianOfnumbers
```

```python
numOfNumbers = myColumn.shape[0]
variance = 0
for i in myColumn:
    variance += (i - meanOfNumbers) ** 2
variance /= 5
variance
```

```python
standartDevation = variance ** 0.5
standartDevation
```

```python
standartError = standartDevation / ((numOfNumbers) ** 0.5)
standartError
```

# Shape Of Distrubition


```python
leftArray = myColumn[:int(myColumn.shape[0]/2)-1].values
rightArray = myColumn[int(myColumn.shape[0]/2):].values
```

```python
Q1 = pd.Series(leftArray).median()
Q3 = pd.Series(rightArray).median()

outLier = Q1 - (1.5 * (Q3 - Q1))
outLier
```
