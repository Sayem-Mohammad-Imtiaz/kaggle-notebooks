```python
def countplt(data, columns=[]):
    for col in columns:
        plt.figure(figsize=(6,6))
        plt.grid()
        sns.countplot(x=col,data=data)
        plt.title(col, fontsize=18)
        plt.show()
def distplt(data, columns=[]):
    for col in columns:
        plt.figure(figsize=(6,6))
        plt.grid()
        sns.distplot(data[col])
        plt.title(col, fontsize=18)
        plt.show()
        
        print('The maximum {} is {}'.format(col,data[col].max()))
        print('The minimum {} is {}'.format(col,data[col].min()))
        print('The average {} is {}'.format(col,data[col].mean()))
        print('With an exceptionally high population of {} {}'.format(col,int(data[col].mode())))

def sctplt(data, base, hue, columns=[]):
    for col in columns:
        plt.figure(figsize=(6,6))
        plt.grid()
        sns.scatterplot(data=df, x=col, y=base,hue=hue)
        plt.title("{} vs {}".format(col,base),fontsize=18)
        plt.show()
```

# EDA Analysis

```python
plt.figure(figsize=(10,10))
plt.grid()
sns.scatterplot(data=df, x="bmi", y="charges")


x= df["bmi"]
y= df["charges"]
m,b = np.polyfit(x,y,1)
plt.plot(x, m*x+b, color='red')
plt.show()
```

```python
insurance_df = df.copy()
insurance_df_smoker=insurance_df[insurance_df['smoker']=='yes']
insurance_df_nonsmoker=insurance_df[insurance_df['smoker']=='no']
# scatterplot
plt.figure(figsize=(8,6))
sns.scatterplot(data=insurance_df_smoker, x="age", y="charges",color='red',label='smoker')
sns.scatterplot(data=insurance_df_nonsmoker, x="age", y="charges",color='blue',label='non-smoker')
plt.title("charges against age", fontsize=15)
plt.plot(insurance_df["age"], 280*insurance_df["age"],color="green")
plt.plot(insurance_df["age"], 280*insurance_df["age"]+22000,color="green")
plt.legend()
plt.show()
```

```python
f, ax = plt.subplots(figsize=(10, 8))
corr = df.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(240,10,as_cmap=True),
            square=True, ax=ax)
```

```python
from sklearn.preprocessing import LabelEncoder
#sex
data = df.copy()
le = LabelEncoder()
le.fit(data.sex.drop_duplicates()) 
data.sex = le.transform(data.sex)
# smoker or not
le.fit(data.smoker.drop_duplicates()) 
data.smoker = le.transform(data.smoker)
#region
le.fit(data.region.drop_duplicates()) 
data.region = le.transform(data.region)
```

```python
f, ax = plt.subplots(figsize=(10, 8))
corr = data.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(240,10,as_cmap=True),
            square=True, ax=ax)
```

It can be easily seen that the charges to be paid depends majorly on three factors:
1. Age
2. BMI
3. If he/she smokes
