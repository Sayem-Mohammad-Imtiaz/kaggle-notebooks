```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv("/kaggle/input/ccdata/CC GENERAL.csv")
print(df.shape)
df.head()
```

```python
df = df.drop(["CUST_ID"],axis=1)
```

```python
df.isna().sum()
```

Filling Empty Values

```python
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].mean(skipna=True), inplace=True)
df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].mean(skipna=True), inplace=True)
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.describe().T
```

From above we can see the outliers but removing them will not resolve any issue.

```python
X = df.copy()
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
X.shape
```

```python
clusters = 25
cost = []
for i in range(1,clusters):
    kmeans = KMeans(i)
    kmeans.fit(X)
    cost.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,8))
plt.plot(cost,'b+-')
plt.show()
```

From the graph we can see that the elbow is comming at the range of 5 to 10. I have chosen 5 as it would be easier to visualize the data. After every increment in the no. of clusters it becomes more and more difficult to visualise after PCA since the dimentionality of the data is very high.

Another thing to keep in mind is that if we want to further improve the clustering model we must try differeent distance measuring techniques.<br><br><br> Some of them are:
1. Classical methods like Euclidean and Manhattan distances
2. Correlation based distance
3. Gower distance
4. Cosine distance

```python
k_means = KMeans(5)
k_means.fit(X)
labels = k_means.labels_
```

```python
clusters = pd.concat([df,pd.DataFrame({"cluster":labels})], axis=1)
```

```python
clusters.head()
```

```python
#5:"purple",6:"black",7:"yellow",8:"pink",
pca = PCA()
principal_components = pca.fit_transform(X)
x,y = principal_components[:,0], principal_components[:,1]

print(principal_components.shape)

colors = {0:"red",1:"blue",2:"green",4:"orange",3:"gray"}
```

```python
final_df = pd.DataFrame({'x': x, 'y':y, 'label':labels}) 
groups = final_df.groupby(labels)
```

```python
fig, ax = plt.subplots(figsize=(15, 10)) 

for name, group in groups:
    ax.plot(group.x, group.y, marker='o', linestyle='', ms=5, color=colors[name], mec='none')
    ax.set_aspect('auto')
    ax.tick_params(axis='x',which='both',bottom='off',top='off',labelbottom='off')
    ax.tick_params(axis= 'y',which='both',left='off',top='off',labelleft='off')
    
ax.set_title("Customers Segmentation based on their Credit Card usage bhaviour.")
plt.show()
```
