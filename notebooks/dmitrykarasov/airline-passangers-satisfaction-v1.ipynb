```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train = pd.read_csv('/kaggle/input/airline-passenger-satisfaction/train.csv', index_col=0)
train.head()
```

```python
train.info()
```

```python
test = pd.read_csv('/kaggle/input/airline-passenger-satisfaction/test.csv', index_col=0)
test.head()
```

```python
test.info()
```

```python
train.dropna(inplace=True)
test.dropna(inplace=True)
```

```python
train.info()
```

```python
test.info()
```

```python
sns.countplot(x='Online boarding', hue="satisfaction", data=train, color="navy")
```

```python
sns.countplot(x='Type of Travel', hue="satisfaction", data=train, color="navy")
```

```python
sns.countplot(x='Class', hue="satisfaction", data=train, color="navy")
```

```python
object_types = train.select_dtypes(include=['object']).columns
object_types
```

```python
train[object_types] = train[object_types].astype('category')
train[object_types] = train[object_types].apply(lambda x: x.cat.codes)
train.head()
```

```python
test[object_types] = test[object_types].astype('category')
test[object_types] = test[object_types].apply(lambda x: x.cat.codes)
test.head()
```

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(20,20))
plt.title('Pearson Correlation', y=1.05, size=15)
sns.heatmap(train.corr(),linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True)

plt.show()
```

```python
colormap = plt.cm.RdBu
plt.figure(figsize=(20,20))
plt.title('Pearson Correlation', y=1.05, size=15)
sns.heatmap(test.corr(),linewidths=0.1,vmax=1.0, 
            square=True, cmap=colormap, linecolor='white', annot=True)

plt.show()
```

Looks like Online boarding, Type of travel and Class features correlate the most with satisfaction

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import recall_score, roc_auc_score, accuracy_score
```

```python
x_train = train.iloc[:,:-1]
x_train.head()
```

```python
x_test = test.iloc[:,:-1]
x_test.head()
```

```python
y_train = train.satisfaction
y_train
```

```python
y_test = test.satisfaction
y_test
```

```python
dtc = DecisionTreeClassifier(random_state=17)
dtc_model = dtc.fit(x_train, y_train)
dtc_predictions = dtc.predict(x_test)
roc_auc_dtc = roc_auc_score(y_test, dtc_predictions)
acc_dtc = accuracy_score(y_test, dtc_predictions)

print(f'ROC AUC Score = {roc_auc_dtc}')
print(f'Accuaracy = {acc_dtc}')
```

```python
nb = GaussianNB()
nb_model = nb.fit(x_train, y_train)
nb_predictions = nb.predict(x_test)
roc_auc_nb = roc_auc_score(y_test, nb_predictions)
acc_nb = accuracy_score(y_test, nb_predictions)

print(f'ROC AUC Score = {roc_auc_nb}')
print(f'Accuaracy = {acc_nb}')
```

```python
rfc = RandomForestClassifier(n_estimators = 100, random_state=17)
rfc_model = rfc.fit(x_train, y_train)
rfc_predictions = rfc.predict(x_test)
roc_auc_rfc = roc_auc_score(y_test, rfc_predictions)
acc_rfc = accuracy_score(y_test, rfc_predictions)

print(f'ROC AUC Score = {roc_auc_rfc}')
print(f'Accuaracy = {acc_rfc}')
```

For prediction Random forest classifier fits better than others
