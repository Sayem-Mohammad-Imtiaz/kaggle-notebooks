```python
df = pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
print(df.shape)
colum_names = df.columns
print(df.columns)
print(df.Outcome.value_counts())
df.head()
```

```python
df.info()
```

```python
y = np.array(df.Outcome) # separa columna Output como labels
y = to_categorical(y) # 1-D array to 2-D array
df = df.drop('Outcome', axis = 1) # Elimina columna Output del dataframe
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size = 0.2, random_state = 20)
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
def define_model():
    model = Sequential()
    model.add(Input(shape=(8, )))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(2, activation='softmax'))

    model.compile(optimizer = 'sgd', loss = 'categorical_crossentropy', metrics = 'accuracy')
    return model
```

```python
model = define_model()
model.summary()
```

```python
history = model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 20, batch_size = 16)
```

```python
target_names = ['No-Diabetes', 'Diabetes']
predictions = model.predict(X_test)
print(classification_report(np.argmax(y_test, axis = 1), np.argmax(predictions, axis = 1), target_names = target_names))
```
