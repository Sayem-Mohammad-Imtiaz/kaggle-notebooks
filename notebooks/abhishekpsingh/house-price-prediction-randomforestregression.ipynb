```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sea
from sklearn.feature_selection import SelectKBest,chi2
```

```python
X = dataset.iloc[:,3:]
y = dataset.iloc[:,2]
```

```python
y.head()
```

```python
plt.scatter(X.sqft_living15,y)
```

```python
X = X.drop(['long'], axis = 1)
```

```python
X_new = SelectKBest(chi2, k=5).fit(X,y)
```

```python
feature_scores = pd.concat([pd.DataFrame(X.columns),pd.DataFrame(X_new.scores_)],axis =1)
```

```python
feature_scores.columns = (['Feature','Score'])
feature_scores
```

```python
print(feature_scores.nlargest(5,'Score'))
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators = 500,max_features = 15,criterion = 'mse',max_depth = 25,min_samples_leaf= 3)
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_train)
```

```python
print(regressor.score(X_train,y_train))
```

```python
print(regressor.score(X_test,y_test))
```

```python
plt.scatter(y_train, y_pred)
plt.xlabel('Real Prices')
plt.ylabel('Predicted Prices')
plt.show()
```
