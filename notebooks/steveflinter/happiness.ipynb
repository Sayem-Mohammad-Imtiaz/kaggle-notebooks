## Data

We start by looking at the 2021 World Happiness Report.

As can be seen by the following table, our data set has a number of columns, containing the different _features_ of our data, and a number of rows, each representing the data from one particular country.

For the columns, for the purposes of this paper, we're only really intersted in a few of them:

* __Country name__ - the name of the country
* __Region indicator__ - the region (continent, or similar) that the country is in
* __Ladder score__ - this is our measure of happiness that a particular country reports
* __Logged GDP per capita__ - this is the GDP per capita of the country, which is a measure of its wealth

We note that both the _Ladder score_ and _Logged GDP per capita_ data elements are shown on a log scale.

First we need to load our data into the notebook.

```python
df = pd.read_csv('/kaggle/input/world-happiness-report-2021/world-happiness-report-2021.csv')
df.head()
```

For the purpose of simplicity, I'm going to rename some of the columns to make them easier to work with.

```python
df[["Country name", "Ladder score", "Logged GDP per capita"]].head(10)
df.rename(columns={"Country name": 'name',
                   "Regional indicator": 'region',
                   "Ladder score": 'happiness',
                   "Logged GDP per capita": 'gdp'}, inplace=True)
```

```python
df_score_gdp = df[["region", "happiness", "gdp"]]
df_score_gdp.head()
```

# Plots

Let's start by plotting the relationship between the GDP score and the happiness score.

```python
sns.set_theme()

# Create a visualization
sns.relplot(
    data = df,
    x="gdp", y="happiness",
    hue="region",
    height=9
)
```

## Initial Analysis

As we can see from the graph above, it does appear that countries with high GDP are highly correlated with high happiness scores. There may be a number of reasons to explain this:

* wealth may cause happiness: perhaps wealthier countries are able to afford services like good healthcare, transportation, education, public services, etc. all of which lead to public happiness
* happiness may drive wealth: it may be that happier people work harder and find better jobs, and hence, lead to countries to become wealthier
* it's also possible that there is some other, unknown reason that links both of these

```python
sns.lmplot(data=df, x="gdp", y="happiness",
          height=10)
```
