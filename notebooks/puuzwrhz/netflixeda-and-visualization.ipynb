```python

imdb_ratings=pd.read_csv('../input/netfliximdb/IMDb ratings.csv',usecols=['weighted_average_vote'])
imdb_titles=pd.read_csv('../input/netfliximdb/IMDb movies.csv', usecols=['title','year','genre'])

ratings = pd.DataFrame({'Title':imdb_titles.title,  'Release Year':imdb_titles.year,'Rating': imdb_ratings.weighted_average_vote,'Genre':imdb_titles.genre})
#usecols=====> select the specific columns from the already read data frame.
ratings.drop_duplicates(subset=['Title','Release Year','Rating'], inplace=True) # Creating new data frame.
ratings.shape
```

```python
ratings
```

```python
ratings.dropna()
joint_data=ratings.merge(df,left_on='Title',right_on='title',how='inner')
joint_data=joint_data.sort_values(by='Rating', ascending=False)
#joint_data.to_csv('../input/netfliximdb/jointdata.csv')
joint_data.shape
```

```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
text = " ".join(review for review in movies_df.description)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
genrtext = " ".join(review for review in movies_df.listed_in)
print ("There are {} words in the combination in the genres.".format(len(genrtext)))
```

```python
stopwords = set(STOPWORDS)
wordcl = WordCloud(stopwords=stopwords, background_color="white", max_font_size=80, max_words= 2000).generate(genrtext)
plt.figure(figsize=(15, 8))
plt.imshow(wordcl, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
mask = np.array(Image.open("../input/netfliximdbimg/3.jpg"))
# The masks  in the following command helps formning the word cloud in a  masked shape of the image provided.
```

```python
wc = WordCloud(background_color='black', mask=mask, mode='RGB', 
              width=1000, max_words=200, height=1000,
              random_state=1)
wc.generate(genrtext)
plt.figure(figsize=(10, 10))
plt.imshow(wc, interpolation='bilinear')
plt.tight_layout(pad=0)
plt.axis('off')
plt.show()
```

```python
wc = WordCloud(background_color='white', mask=mask, mode='RGB',
               width=1000, max_words=1000, height=1000,
               random_state=1, contour_width=1, 
               contour_color='steelblue')
wc.generate(genrtext)
plt.figure(figsize=(10, 10))
plt.imshow(wc, interpolation='bilinear')
plt.tight_layout(pad=0)
plt.axis('off')
plt.show()
```
