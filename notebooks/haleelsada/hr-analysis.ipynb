```python
data=pd.read_csv('/kaggle/input/hr-analytics/HR_comma_sep.csv')
data.head()
```

```python
data.columns
```

```python
# Delete number_project,promotion_last_5year and Work_accident
# label encode last two columns
from sklearn.preprocessing import LabelEncoder as LE
le=LE()
data=data.drop(['number_project','promotion_last_5years', 'Work_accident'],axis=1)
data['Department']=le.fit_transform(data['Department'])
data['salary']=le.fit_transform(data['salary'])
```

```python
data.head()
```

```python
data.corr()
```

It looks like satisfaction is one of the reasons why people are leaving

```python
# check how satisfied is people who left and don't left
import seaborn as sbn
import matplotlib.pyplot as plt
sbn.histplot(x=data[data.left==1].satisfaction_level,kde=False,color='red')
sbn.histplot(x=data[data.left==0].satisfaction_level,kde=False)
plt.legend()
```

conclusion:satisfaction is one of the main reason for people leaving the job

```python
# make a contengency table of left and salary
cttable = pd.crosstab(data['left'],
                            data['salary'], 
                               margins = False)

cttable.loc[len(cttable.index)] = [sum(cttable[0]),sum(cttable[1]),sum(cttable[2])] 
cttable
```

```python
# make the contengency table column frequency contengency talble
# here 0=left,1=not left and 0=high 1=low 2=medium
cttable.loc[0] = cttable.loc[0]/cttable.loc[2]
cttable.loc[1] = cttable.loc[1]/cttable.loc[2]
cttable.iloc[:2]
```

```python
# plot it in a heatmap to see how leaving the job is related with salary 
sbn.heatmap(cttable.iloc[:2])
plt.text(0.2, 0.5, 'high')#'0= high 1= low 2=med')
plt.text(1.5, 0.5, 'low')
plt.text(2.3, 0.5, 'med')
```

It seems people with high icome tends to leave their job than low income,what a contradiction!

```python
cttable2 = pd.crosstab(data['left'],
                            data['Department'], 
                               margins = False)

cttable2.loc[2] = [sum(cttable2[0]),sum(cttable2[1]),sum(cttable2[2]),
                                    sum(cttable2[3]),sum(cttable2[4]),sum(cttable2[5]),
                   sum(cttable2[6]),sum(cttable2[7]),sum(cttable2[8]),sum(cttable2[9])] 
for i in range(2):
    
    cttable2.loc[i] = cttable2.loc[i]/cttable2.loc[2]

cttable2.iloc[:2]
```

```python
sbn.heatmap(cttable2.iloc[:2])
```

```python
from sklearn.model_selection import train_test_split as tts
#ft=data.drop(['left'],axis=1)
ft=data[['Department','satisfaction_level','salary']]
label=data.left
X,x,Y,y=tts(ft,label)
```

Taking some features is better than taking all

```python
# simple model
from sklearn.linear_model import LogisticRegression as lr
```

```python
# model training and find accuracy
model=lr()
model.fit(X,Y)
model.score(x,y)
```
