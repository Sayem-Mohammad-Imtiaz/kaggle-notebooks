```python
#importing all modules
import numpy as np
import pandas as pd
import scipy as sp
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import os
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
#reading train data 
df=pd.read_csv("/kaggle/input/salary-data-simple-linear-regression/Salary_Data.csv")
df.dropna(inplace=True)
x=np.array(df["YearsExperience"]).reshape(-1, 1)
y=np.array(df["Salary"]).reshape(-1, 1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33)
reg=LinearRegression().fit(x_train, y_train)
```

```python
#training moddle
reg=LinearRegression(normalize=True).fit(x_train, y_train)
y_train_pred=reg.predict(x_train)
print("Training Score -" ,r2_score(y_train,y_train_pred))
```

```python
#Train Data Visualization
%matplotlib inline
fig = plt.figure()
plt.plot(x_train,y_train_pred,color='black')
plt.scatter(x_train,y_train,color='red',linewidth=0.1)
fig.set_figheight(6)
fig.set_figwidth(20)
plt.show()
```

```python
#reading test data 
y_pred = reg.predict(x_test)
print("Prediction Score -",reg.score(x_test,y_test))
```

```python
#Test Data Visualization
%matplotlib inline
fig = plt.figure()
plt.plot(x_test,y_pred,color='black')
plt.scatter(x_test,y_test,color='red',linewidth=0.1)
fig.set_figheight(6)
fig.set_figwidth(20)
plt.show()
```
