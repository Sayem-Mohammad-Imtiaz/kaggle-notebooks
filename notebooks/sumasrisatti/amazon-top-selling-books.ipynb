```python
plt.subplot(221)

df['Reviews'].value_counts().plot(kind='bar', title='Reviews', figsize=(16,9),color = 'green')

plt.xticks(rotation=0)

plt.subplot(222)

df['Price'].value_counts().plot(kind='bar', title='Price',color = 'pink')

plt.xticks(rotation=0)

plt.subplot(223)

df['Year'].value_counts().plot(kind='bar', title='Year',color='violet')

plt.xticks(rotation=0)

plt.subplot(224)

df['Genre'].value_counts().plot(kind='bar', title='Genre',color='orange')

plt.xticks(rotation=0)

plt.show()
```

observation: average review for all the books is between 8 to 10. 8 is the most common book price.sales rate of non fiction books are more than fictional books

here the subplot(221) means it consists of 2 rows and 2 columns.
we use figsize attritube to specify the width and height of a figure.
if required title can be given to the plot which must be written in ' ' i.e title = ' '

```python
df.boxplot()
```

boxplot displays the distribution of data on basis of minimum,maximum,median,first quartile and second quartile.it also indicates whether the data is symmetrical are not.

```python
sns.heatmap(df.isnull())
```

this plots the missing values on a heatmap using seaborn

```python
sns.distplot(df['Price'])
```

observation: price between 0 to 20 has highest density

 distplot gives as histogram with a line over it.distplot can be used to represent the univariate distribution of data. 

```python
corr = df.corr()
sns.heatmap(corr, annot=True, square=True)
plt.yticks(rotation=2)
plt.show()
```

annot = True makes sure that the values of each cell appear on the chart and yticks sets the y-axis values

```python
df.hist(figsize=(10, 12), bins=50, xlabelsize=8, ylabelsize=8); 
```

this plots a histogram with width as 10 and height as 12.bins represents the bars of histogram.bins can be changed as per our wish

```python
sns.relplot(x="User Rating", y="Price", hue='Genre', data=df)
```

observation:the plot is more densed in range 4.2 to 5 below the price 60.the nonfiction book with 4.4 rating have the highest price

relplot in seaborn is used for visualixing statistical relationship with scatter plot and line plot. x and y axes can be given any names required for plotting.hue represents the column in which we want to use color encoding 

```python
sns.factorplot("Reviews",data=df,kind="count")
```

```python
sns.factorplot("Price",data=df,kind="count")
```

observation: the price of 8 rupees have highest count

factorplot is used to represent different types of categorical plot.the type of plot can be specified using 'kind' parameter

```python
sns.set_palette("gist_rainbow_r")
sns.jointplot(x="Price",y="User Rating",kind="hex",data=df)
plt.show
```

important obervations from the dataset:nonfiction books are sold more than fiction books.maximum of the books are of price 8.the highest priced book was rated between 4.4 and 4.6.almost all of the books are rated more than 4

jointplot visualizes the relationship between two variables and describe their individual distributions on the sameplot and kind='hex' gives the plotting a hexagonal shape

points to be noted for good data visualization:
1.information have to be simple 
2.headings must be clear so that the reader can understand as soon the person looks at the graph
3.graph must be choosen based on the information and the message to be conveyed
4.3d visualization can be misleading so it should be avoided
