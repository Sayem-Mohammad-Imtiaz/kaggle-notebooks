```python
def square():
    
    def add():
        x=2
        y=3
        z=x+y
        return z
    
    
    return add()**2
print(square())
```

 DEDFAULT and FLEXILBE ARGUMENT
 
 default argüman example:
 def f(a,b=1):  # b burda default argümandır
 
 flexible argüman:
 def f(*args):    # *args ile bir ve birden fazla değişken kullanmamıza yarar
 
 def f(*kwargs):  # bir sözlük argümanıdır
 
 


```python
def f(a,b=2,c=3):  #default argümanlar tanımladık
    y = a+b+c    
    return y
print(f(5))  # a dan başka diğer değişkenlerin parametersi oldğu için , sadece a değişkenine atadık 
 
print(f(5,3,6))  # default değerleride değiştirdik
```

```python
#flexible argüman

def f(*args):
    for i in args:
        print(i)
f(1)
print(" ")
f(1,2,3,4,5,6)
print(" ")

#flexible argüman **kwargs 
def f(**kwargs):
    for key,value in kwargs.items():
        print(key,":",value)
        
f(country = "spain",capital="madrid",population=1234534)
```
