# LINE PLOT

```python
#now we will visualize Turkey's unviersities' yearly citations and teaching values
#first we will import graph objects as "go"
import plotly.graph_objs as go

#secondly we need to define traces
trace1 = go.Scatter(
                    x = Turkeydf.year,
                    y = Turkeydf.citations,
                    mode = "lines",
                    name = "citations",
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    text = Turkeydf.university_name)

trace2 = go.Scatter(
                    x = Turkeydf.year,
                    y = Turkeydf.teaching,
                    mode = "lines+markers",
                    name = "teaching",
                    marker = dict(color = 'rgba(80, 26, 80, 0.8)'),
                    text= Turkeydf.university_name)

#thirdly we will make a list by combining both traces
data = [trace1, trace2]

#then we will make a layout for what we want to see outside of figure
layout = dict(title = "Citation and Teaching by year of Turkey's Universities",
             xaxis = dict(title = 'World Rank', ticklen = 5, zeroline = False))

#now we will make our figure
fig = dict(data = data, layout = layout)

#lastly we will plot our figure
iplot(fig)
```

# SUBPLOTS

```python
trace1 = go.Scatter(
    x=Turkeydf.year,
    y=Turkeydf.research,
    name = "research",
    marker = dict(color = 'rgba(255, 112, 0, 0.8)'),
    text = Turkeydf.university_name   
)
trace2 = go.Scatter(
    x=Turkeydf.year,
    y=Turkeydf.citations,
    xaxis='x2',
    yaxis='y2',
    name = "citations",
    marker = dict(color = 'rgba(0, 255, 122, 0.8)'),
    text = Turkeydf.university_name   
)
trace3 = go.Scatter(
    x=Turkeydf.year,
    y=Turkeydf.income,
    xaxis='x3',
    yaxis='y3',
    name = "income",
    marker = dict(color = 'rgba(122, 0, 122, 0.8)'),
    text = Turkeydf.university_name
)
trace4 = go.Scatter(
    x=Turkeydf.year,
    y=Turkeydf.total_score,
    xaxis='x4',
    yaxis='y4',
    name = "total_score",
    marker = dict(color = 'rgba(0, 122, 255, 0.8)'),
    text = Turkeydf.university_name
)
data = [trace1, trace2, trace3, trace4]
layout = go.Layout(
    xaxis=dict(
        domain=[0, 0.45]
    ),
    yaxis=dict(
        domain=[0, 0.45]
    ),
    xaxis2=dict(
        domain=[0.55, 1],
        anchor='x2'
    ),
    xaxis3=dict(
        domain=[0, 0.45],
        anchor='x3'
    ),
    xaxis4=dict(
        domain=[0.55, 1],
        anchor='x4'
    ),
    yaxis2=dict(
        domain=[0, 0.45],
        anchor='y2'
    ),
    yaxis3=dict(
        domain=[0.55, 1],
        anchor='y3'
    ),
    yaxis4=dict(
        domain=[0.55, 1],
        anchor='y4'
    ),
    title = "Research, citation, income and total score by year of Turkey's Universities"
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

# BUBBLE PLOT

```python
# data preparation
df2016 = timesData[timesData.year == 2016].iloc[:20,:]
num_students_size  = [float(each.replace(',', '.')) for each in df2016.num_students]
international_color = [float(each) for each in df2016.international]
data = [
    {
        'y': df2016.teaching,
        'x': df2016.world_rank,
        'mode': 'markers',
        'marker': {
            'color': international_color,
            'size': num_students_size,
            'showscale': True   
        },
        "text" :  df2016.university_name    
    }
]
iplot(data)
# in here we see that if bubbles are bigger that means num_students are bigger, if colos are more yellow means international score is higher
```

# PIE PLOT

```python
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
trace = go.Pie(
            labels = langs, 
            values = students,         
            textinfo='label+percent',   #in here we write label and percent on it together
            hoverinfo='label+percent',   #in here we see what it will be seen when we move mouse on
            hole = .3,                   #in here we make a hole and .3 hole, .7 graphic
            pull = [0, 0, 0.2, 0, 0],     #in here we take out the 3rd value means Java
            )
data = [trace]

layout = dict(title = "Languages vs Student")
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```
