```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

def eval_metrics(y_set,y_pred):
    MAE = mean_absolute_error(y_set, y_pred)
    MSE = mean_squared_error(y_set, y_pred)
    r2 = r2_score(y_set, y_pred)
    values = {'MAE': round(MAE,3), 'MSE' : round(MSE,3), 'r2' : round(r2,3)}
    print(f'R2_score: {r2 * 100:.2f}%')
    print(f'mean absolute error: {MAE:.2f}')
    print(f'mean squared error: {MSE:.2f}')

    return values
```

```python
X = data_set[['pretest','school_setting','school_type','teaching_method','n_student','lunch']]
Y = data_set[['posttest']]
X = pd.get_dummies(X)
X.head(100)
```

```python
[x_adj, x_tst, y_adj, y_tst] = train_test_split(X,Y,test_size=0.45, random_state=5)

model_linear= LinearRegression()
model_linear.fit(x_adj, y_adj)

y_pred = model_linear.predict(x_tst)

multi_eval = eval_metrics(y_tst,y_pred)
```
