# Data Loading

```python
df = pd.read_csv('/kaggle/input/singapore-airbnb/listings.csv')
# df.sort_values(by='minimum_nights')
df.head()
```

# Manual Mean and Variance Calculation

```python
m_id = sum(df['id'])/len(df['id'])
m_host_id = sum(df['host_id'])/len(df['host_id'])
m_lat = sum(df['latitude'])/len(df['latitude'])
m_long = sum(df['longitude'])/len(df['longitude'])
m_price = sum(df['price'])/len(df['price'])
m_min_night = sum(df['minimum_nights'])/len(df['minimum_nights'])
m_num_review = sum(df['number_of_reviews'])/len(df['number_of_reviews'])
m_reviews_per_month = sum(df['reviews_per_month'])/len(df['reviews_per_month'])
m_cal_host_list_count = sum(df['calculated_host_listings_count'])/len(df['calculated_host_listings_count'])
m_avail = sum(df['availability_365'])/len(df['availability_365'])


print('Manual Mean Calculation\n'+('='*25))
print(f'id: {m_id}\nhost_id: {m_host_id}\nlatitude: {m_lat}\nlongitude: {m_long}')
print(f'price: {m_price}\nminimung nights: {m_min_night}\nnumber of reviews: {m_num_review}')
print(f'reviews per month: {m_reviews_per_month}\ncalculated hsot listings count: {m_cal_host_list_count}\navailable 365: {m_avail}')
```

```python
v_id = sum((df['id']-m_id)**2)/(len(df['id'])-1)
v_host_id = sum((df['host_id']-m_host_id)**2)/(len(df['host_id'])-1)
v_lat = sum((df['latitude']-m_lat)**2)/(len(df['latitude'])-1)
v_long = sum((df['longitude']-m_long)**2)/(len(df['longitude'])-1)
v_price = sum((df['price']-m_price)**2)/(len(df['price'])-1)
v_min_night = sum((df['minimum_nights']-m_min_night)**2)/(len(df['minimum_nights'])-1)
v_num_review = sum((df['number_of_reviews']-m_num_review)**2)/(len(df['number_of_reviews'])-1)
v_reviews_per_month = sum((df['reviews_per_month']-m_reviews_per_month)**2)/(len(df['reviews_per_month'])-1)
v_cal_host_list_count = sum((df['calculated_host_listings_count']-m_cal_host_list_count)**2)/(len(df['calculated_host_listings_count'])-1)
v_avail = sum((df['availability_365']-m_avail)**2)/(len(df['availability_365'])-1)


print('Manual Variance Calculation\n'+('='*25))
print(f'id: {v_id}\nhost_id: {v_host_id}\nlatitude: {v_lat}\nlongitude: {v_long}')
print(f'price: {v_price}\nminimung nights: {v_min_night}\nnumber of reviews: {v_num_review}')
print(f'reviews per month: {v_reviews_per_month}\ncalculated host listings count: {v_cal_host_list_count}\navailable 365: {v_avail}')
```
