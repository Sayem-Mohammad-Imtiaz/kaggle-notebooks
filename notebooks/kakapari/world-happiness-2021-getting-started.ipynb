### What is Ladder Score?

The **Cantril ladder**: it asks respondents to think of a ladder, with the best possible life for them being a 10, and the worst possible life being a 0. They are then asked to rate their own current lives on that 0 to 10 scale. 

```python
fig, ax = plt.subplots(1,1, figsize=(15, 8), dpi=100)
L = ax.legend(frameon=False,loc="upper center", bbox_to_anchor=(1.0, 0.8), ncol= 1)
ax = sns.kdeplot(df['Ladder score'], hue=df['Regional indicator'], fill=True, linewidth=2)
```
