# **Distribuição de palavras por Rewiew**

```python
df = pd.read_csv('../input/trip-advisor-hotel-reviews/tripadvisor_hotel_reviews.csv')
df.head()
df.drop(['Rating'], axis=1, inplace=True)
df.shape
```

Remoção

```python
df['Review'] = df['Review'].apply(lambda x: x.translate(str.maketrans('', '', string.punctuation))) #removendo a pontuação
df['review_no_stopw'] = df['Review'].apply(lambda x: remove_stopwords(x)) #para a remoção
df.shape
```

```python
df.iloc[0]['Review']
```

*Contagem de palavras**

```python
df['num_words'] = df['Review'].apply(lambda x: len(x.split()))
df['num_words_no_stopw'] = df['review_no_stopw'].apply(lambda x: len(x.split()))
df.shape
```
