# Line Charts

Line Charts Example: Economy and Freedom vs Happiness_Rank of Top 157 Countries in 2015


```python
# prepare data frame
data_2015 = data_2015.iloc[:158,:]


# Creating trace1
trace1 = go.Scatter(
                    x = data_2015.Happiness_Rank,
                    y = data_2015.Economy,
                    mode = "lines",
                    name = "Economy",
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    text= data_2015.Country)
# Creating trace2
trace2 = go.Scatter(
                    x = data_2015.Happiness_Rank,
                    y = data_2015.Freedom,
                    mode = "lines+markers",
                    name = "Freedom",
                    marker = dict(color = 'rgba(80, 26, 80, 0.8)'),
                    text= data_2015.Country)
data = [trace1, trace2]
layout = dict(title = 'Economy and Freedom vs Happiness Rank of Top 157 Countries',
              xaxis= dict(title= 'Happiness Rank',ticklen= 5,zeroline= False)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```

# Scatter Plot

```python


df_WE = data_2015[data_2015.Region == 'Western Europe'].iloc[:158,:]
df_NA = data_2015[data_2015.Region == 'North America'].iloc[:158,:]
df_EA = data_2015[data_2015.Region == 'Eastern Asia'].iloc[:158,:]

# creating trace1
trace1 =go.Scatter(
                    x = df_WE.Happiness_Rank,
                    y = df_WE.Trust,
                    mode = "markers",
                    name = "Western Europe",
                    marker = dict(color = 'rgba(255, 128, 255, 0.8)'),
                    text= df_WE.Country)
# creating trace2
trace2 =go.Scatter(
                    x = df_NA.Happiness_Rank,
                    y = df_NA.Trust,
                    mode = "markers",
                    name = "North America",
                    marker = dict(color = 'rgba(255, 128, 2, 0.8)'),
                    text= df_NA.Country)
# creating trace3
trace3 =go.Scatter(
                    x = df_EA.Happiness_Rank,
                    y = df_EA.Trust,
                    mode = "markers",
                    name = "Eastern Asia",
                    marker = dict(color = 'rgba(0, 255, 200, 0.8)'),
                    text= df_EA.Country)
data = [trace1, trace2, trace3]
layout = dict(title = 'Happiness Rank vs Trust of top 157 Countries with 3 regions',
              xaxis= dict(title= 'Happiness Rank',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'Trust',ticklen= 5,zeroline= False)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```

# Histogram

Lets look at histogram of Generosity ratio in Western Europe and Eastern Asia

```python

# prepare data
WesternEurope= data_2015.Generosity[data_2015.Region == 'Western Europe']
EasternAsia = data_2015.Generosity[data_2015.Region == 'Eastern Asia']

trace1 = go.Histogram(
    x=WesternEurope,
    opacity=0.75,
    name = "Western Europe",
    marker=dict(color='rgba(171, 50, 96, 0.6)'))
trace2 = go.Histogram(
    x=EasternAsia,
    opacity=0.75,
    name = "Eastern Asia",
    marker=dict(color='rgba(12, 50, 196, 0.6)'))

data = [trace1, trace2]
layout = go.Layout(barmode='overlay',
                   title=' Generosity in Western Europe and Eastern Asia',
                   xaxis=dict(title='Gnerosity'),
                   yaxis=dict( title='Count'),
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

# Scatter Matrix Plots

Scatter Matrix = it helps us to see covariance and relation between more than 2 features
Example Scatter Matrix Plots: Correlation analysis with Happiness Score, Economy, Freedom and Trust.

```python
# import figure factory

import plotly.figure_factory as ff


# prepare data

d2015 = data_2015.loc[:,["Happiness_Score","Economy", "Freedom","Trust"]]
d2015["index"] = np.arange(1,len(d2015)+1)

# scatter matrix
fig = ff.create_scatterplotmatrix(d2015, diag='box', index='index',colormap='Portland',
                                  colormap_type='cat',
                                  height=700, width=700)
iplot(fig)
```

# Inset Plots

Inset Matrix = 2 plots are in one frame

```python
# first line plot
trace1 = go.Scatter(
    x=data_2015.Happiness_Rank,
    y=data_2015.Freedom,
    name = "Freedom",
    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
)
# second line plot
trace2 = go.Scatter(
    x=data_2015.Happiness_Rank,
    y=data_2015.Economy,
    xaxis='x2',
    yaxis='y2',
    name = "Economy",
    marker = dict(color = 'rgba(160, 112, 20, 0.8)'),
)
data = [trace1, trace2]
layout = go.Layout(
    xaxis2=dict(
        domain=[0.6, 0.95],
        anchor='y2',        
    ),
    yaxis2=dict(
        domain=[0.6, 0.95],
        anchor='x2',
    ),
    title = 'Feedom and Economy vs Happiness Rank of Countiries'

)

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

# 3D Scatter Plot with Colorscaling

3D Scatter: Sometimes 2D is not enough to understand data. Therefore adding one more dimension increase the intelligibility of the data. Even we will add color that is actually 4th dimension.

Example of 3D Scatter Plot with Colorscaling. Happiness Rank, Family and Health

```python
# create trace 1 that is 3d scatter
trace1 = go.Scatter3d(
    x=data_2015.Happiness_Rank,
    y=data_2015.Family,
    z=data_2015.Health,
    mode='markers',
    marker=dict(
        size=10,
        color='rgb(255,0,0)',                # set color to an array/list of desired values      
    )
)

data = [trace1]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0  
    )
    
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

# Multiple Subplots

Multiple Subplots: While comparing more than one features, multiple subplots can be useful.

Multiple Subplots Example: Happiness_Score, Trust, Freedom and Economy VS World Rank of Countires

```python
trace1 = go.Scatter(
    x=data_2015.Happiness_Rank,
    y=data_2015.Happiness_Score,
    name = "Happiness Score"
)
trace2 = go.Scatter(
    x=data_2015.Happiness_Rank,
    y=data_2015.Trust,
    xaxis='x2',
    yaxis='y2',
    name = "Trust"
)
trace3 = go.Scatter(
    x=data_2015.Happiness_Score,
    y=data_2015.Freedom,
    xaxis='x3',
    yaxis='y3',
    name = "Freedom"
)
trace4 = go.Scatter(
    x=data_2015.Happiness_Rank,
    y=data_2015.Economy,
    xaxis='x4',
    yaxis='y4',
    name = "Economy"
)
data = [trace1, trace2, trace3, trace4]
layout = go.Layout(
    xaxis=dict(
        domain=[0, 0.45]
    ),
    yaxis=dict(
        domain=[0, 0.45]
    ),
    xaxis2=dict(
        domain=[0.55, 1]
    ),
    xaxis3=dict(
        domain=[0, 0.45],
        anchor='y3'
    ),
    xaxis4=dict(
        domain=[0.55, 1],
        anchor='y4'
    ),
    yaxis2=dict(
        domain=[0, 0.45],
        anchor='x2'
    ),
    yaxis3=dict(
        domain=[0.55, 1]
    ),
    yaxis4=dict(
        domain=[0.55, 1],
        anchor='x4'
    ),
    title = 'Happiness_Score, Trust, Freedom and Economy VS World Rank of Countires'
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
