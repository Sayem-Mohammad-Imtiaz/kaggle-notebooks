```python
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter
#
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import plot_confusion_matrix, classification_report
#
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Importando o csv
df = pd.read_csv(r'/kaggle/input/pima-indians-diabetes-database/diabetes.csv')
#Verificando nulos
print(df.isna().sum())
#Amostra do Dataset
df.sample(10)
```

```python
#Plota distribuição dos dados
fig, axe =  plt.subplots(nrows=1,ncols=8, figsize=(30,4))
plt.subplots_adjust(wspace= 0.2)
for i in range(8):
    axe[i].hist(df.iloc[:,i])
    axe[i].grid(linestyle=':', linewidth=0.2)
    tit = df.columns[i]
    axe[i].title.set_text(tit)
    #
    #Printa dados básicos da feature
    print(df[tit].describe())
    print('--------------------------------------------')
    print(' ')
plt.show()
```

```python
#Box plot da idade com anotações de média mediana e max
plt.rcParams['figure.figsize'] = (6,7)
plt.boxplot(df.Age)
median = round(df.Age.median(),2)
mean = round(df.Age.mean(),2)

plt.annotate(('Mediana: %s' %median), 
              xy=(1, median), 
              xytext=(0.75, median + 4),
           arrowprops=
                dict(facecolor='blue', shrink=0.01),
                horizontalalignment='right',
                verticalalignment='center')

plt.annotate(('Média: %s' %mean), 
              xy=(1, mean), 
              xytext=(1.5, mean + 4),
           arrowprops=
                dict(facecolor='blue', shrink=0.05),
                horizontalalignment='right',
                verticalalignment='center')


plt.annotate(('Maior Valor: %s' %round(max(df.Age),2)), 
              xy=(1, max(df.Age)), 
              xytext=(1.5, max(df.Age)),
           arrowprops=
                dict(facecolor='blue', shrink=0.05),
                horizontalalignment='right',
                verticalalignment='center')
#
plt.title('Idade')
plt.xticks([])
plt.show()
```

```python
#Correlação (Pearson) entre as features
corr_diabetes = df.corr()
img = corr_diabetes.style.background_gradient(cmap='coolwarm')
img
```

```python
y = df.Outcome
X = df[['Pregnancies','Glucose','BloodPressure','SkinThickness','Insulin','BMI','DiabetesPedigreeFunction','Age']]
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
#Treina com X_train
model = LogisticRegression(max_iter=200)
model.fit(X_train,y_train)
#relevância das features
importance = model.coef_[0]

# plota coef
#
plt.bar([x for x in range(len(importance))], importance)
plt.grid(linestyle=':', linewidth=0.5)
plt.title('Coef')
plt.show()
```

```python
#Score com dados de treino
score_train = (model.score(X_train,y_train))
print('Score com dados de treino : %f' %score_train)
```

```python
# predict
predict_y = model.predict(X_test)
#
# Mean squared error
print('Mean squared error: ', round(mean_squared_error(y_test, predict_y),2))
#Coeficiente de determinação - 1 é perfeito
print('Coeficiente de determinação: ', round(r2_score(y_test, predict_y),2))
```

```python
#Matriz de confusão com dados de teste
plot_confusion_matrix(model,X_test,y_test);
```

```python
print(classification_report(y_test,model.predict(X_test)))
```
