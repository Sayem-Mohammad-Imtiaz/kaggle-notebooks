# DATA AUGMENATION

### Class Imbalance or data scarcity problem, data augmentation got solution. It is a way of mutilplying images by different image transforms and preprocessing.There are four modules giving four ways to implement data augmentation!

## The first module is to implement data augmentation using tensorflow and keras

## IMPORT LIBRARIES AND DEFINE NECESSARY FUNCTIONS



```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from matplotlib.pyplot import imread,imshow, subplots, show

def plot(data_generator):
    
    """
    plots 4 images generated by the ImageDataGenerator
    """
    data_generator.fit(images)
    image_iterator = data_generator.flow(images)
    
    # Plot the images in iteration
    fig, rows = subplots(nrows=1, ncols=4, figsize=(18,18))
    for row in rows:
        row.imshow(image_iterator.next()[0])#.astype('int')
        row.axis('off')
    show()
```

## LOAD THE DATA AND READ THE IMAGES

```python
image = imread("../input/pokemon-images-and-types/images/images/abomasnow.png")
images = image.reshape(1, image.shape[0], image.shape[1], image.shape[2])
imshow(images[0])
show()
```

## ROTATION:

```python
#1 - rotation from +90 to -90 (can change the parameter too)
data_generator = ImageDataGenerator(rotation_range=90)
plot(data_generator)
```

## HORIZONTAL SHIFT:

```python
# 2 Horizontal shifting
data_generator = ImageDataGenerator(width_shift_range=0.3)
plot(data_generator)
```

## VERTICAL SHIFT

```python
# 3 vertical shifting
data_generator = ImageDataGenerator(height_shift_range=0.3)
plot(data_generator)
```

## INTENSITY MODIFICATION:

```python
# 4 Intensity and contrast # change the astype to int in the plot function
data_generator = ImageDataGenerator(brightness_range=(0.1, 0.9))
plot(data_generator)
```

## SHEARING:

```python
# 5 shearing and intensity
data_generator = ImageDataGenerator(shear_range=45.0)
plot(data_generator)
```

## ZOOMING:

```python
# 6 zoom
data_generator = ImageDataGenerator(zoom_range=[0.5, 1.5])
plot(data_generator)
```

## HORIZONTAL FLIP:

```python
# 7 Horizontal flip
data_generator = ImageDataGenerator(horizontal_flip=True)
plot(data_generator)
```

## VERTICAL FLIP:

```python
# 8 vertical flip
data_generator = ImageDataGenerator(vertical_flip=True)
plot(data_generator)
```

## Using tensorflow

```python
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
import tensorflow_datasets as tfds
import tensorflow.keras as layers
```

```python
(train_ds, val_ds, test_ds), metadata = tfds.load(
'tf_flowers',
split=['train[:80%]', 'train[80%:90%]', 'train[90%:]'],
with_info=True,
as_supervised=True,
)
```

```python
get_label_name = metadata.features['label'].int2str
image, label = next(iter(train_ds))
_ = plt.imshow(image)
_ = plt.title(get_label_name(label))
```

```python
def visualize(original, augmented):
    fig = plt.figure()
    plt.subplot(1,2,1)
    plt.title('Original image')
    plt.imshow(original)
    plt.subplot(1,2,2)
    plt.title('Augmented image')
    plt.imshow(augmented)
```

```python
flipped = tf.image.flip_left_right(image)
visualize(image, flipped)
```

```python
grayscaled = tf.image.rgb_to_grayscale(image)
visualize(image, tf.squeeze(grayscaled))
_ = plt.colorbar()
```

```python
saturated = tf.image.adjust_saturation(image, 3)
visualize(image, saturated)
```

```python
bright = tf.image.adjust_brightness(image, 0.4)
visualize(image, bright)
```

```python
cropped = tf.image.central_crop(image, central_fraction=0.5)
visualize(image,cropped)
```

```python
rotated = tf.image.rot90(image)
visualize(image, rotated)
```

## Using skimage
https://medium.com/analytics-vidhya/image-augmentation-for-deep-learning-using-pytorch-feature-engineering-for-images-3f4a64122614


```python
# importing all the required libraries
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import skimage.io as io
from skimage.transform import rotate, AffineTransform, warp
from skimage.util import random_noise
from skimage.filters import gaussian
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading the image using its path
image = io.imread('../input/pokemon-images-and-types/images/images/abomasnow.png')
# shape of the image
print(image.shape)
# displaying the image
io.imshow(image)
```

```python
print('Rotated Image')
#rotating the image by 45 degrees
rotated = rotate(image, angle=45, mode = 'wrap')
#plot the rotated image
io.imshow(rotated)
```

```python
#flip image left-to-right
flipLR = np.fliplr(image)
plt.imshow(flipLR)
plt.title('Left to Right Flipped')
```

```python
#flip image up-to-down
flipUD = np.flipud(image)
plt.imshow(flipUD)
plt.title('Up Down Flipped')
```

```python
#standard deviation for noise to be added in the image
sigma=0.155
#add random noise to the image
noisyRandom = random_noise(image,var=sigma**2)
plt.imshow(noisyRandom)
plt.title('Random Noise')
```

```python
#blur the image
blurred = gaussian(image,sigma=1,multichannel=True)
plt.imshow(blurred)
plt.title('Blurred Image')
```

##  Using Pytorch

```python
import os 
import torch  
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image

import torchvision

%matplotlib inline
```

```python
# you can use any image of your choice
img = Image.open('../input/pokemon-images-and-types/images/images/abomasnow.png')
plt.imshow(img)
```

```python

center_crop = torchvision.transforms.CenterCrop(size=(200,300))
img = center_crop(img)
plt.imshow(img)
```

```python

color_jitter = torchvision.transforms.ColorJitter(brightness=0.5, contrast=0.5, saturation=0.5, hue=0.5)
img = color_jitter(img)
plt.imshow(img)
```

```python
gray = torchvision.transforms.Grayscale(num_output_channels=1)
img = gray(img)
plt.imshow(img, cmap='gray')
```

```python
pad = torchvision.transforms.Pad(50, fill=0, padding_mode='constant')
img = pad(img)
plt.imshow(img)
```

```python
random_crop = torchvision.transforms.RandomCrop((200, 300), fill=0, padding_mode='constant')
img = random_crop(img)
plt.imshow(img)
```

```python
horizontal_flip = torchvision.transforms.RandomHorizontalFlip(p=1)
img = horizontal_flip(img)
plt.imshow(img)
```

```python
vertical_flip = torchvision.transforms.RandomVerticalFlip(p=1)
img = vertical_flip(img)
plt.imshow(img)
```

```python
random_persp = torchvision.transforms.RandomPerspective(distortion_scale=0.5, p=1, interpolation=3, fill=0)
img = random_persp(img)
plt.imshow(img)
```

```python
random_rotation = torchvision.transforms.RandomRotation(degrees = 45)
img = random_rotation(img)
plt.imshow(img)
```

```python
tensor = torchvision.transforms.ToTensor()
random_erase = torchvision.transforms.RandomErasing(p=1)

img = tensor(img)
img = random_erase(img)
plt.imshow(img.permute(1,2,0))
```

```python
# All combined

img = Image.open('../input/pokemon-images-and-types/images/images/abomasnow.png')

transforms = torchvision.transforms.Compose([
    torchvision.transforms.Resize((300,400)),
    torchvision.transforms.ColorJitter(hue=.05, saturation=.05),
    torchvision.transforms.RandomHorizontalFlip(),
    torchvision.transforms.RandomRotation(20)
])

img = transforms(img)
plt.imshow(img)
```
