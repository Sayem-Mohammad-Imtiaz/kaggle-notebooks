```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/kaggle/input/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv")
```

# Removing Active Ratio, Discharge Ratio and Death Ratio since they can be extracted from their respective columns

```python
data.drop(columns=["Active", "Discharged", "Deaths"], axis=1, inplace=True)
data.rename(columns={"Active Ratio (%)":"Active",
                    "Discharge Ratio (%)":"Discharge",
                    "Death Ratio (%)":"Death"}, inplace=True)

data.Active = data.Active/100
data.Death = data.Death/100
data.Discharge = data.Discharge/100
```

```python
data.set_index("State/UTs", inplace=True)
```

# Arrange states by their Total Cases

```python
data.sort_values(by="Total Cases", ascending=False)["Total Cases"].plot.bar(figsize=(15, 5), color="b", ylabel="Total Counts", title="Total Cases")
plt.axhline(y=data["Total Cases"].mean(), color="r", linestyle="-")
plt.text(x=29, y=data["Total Cases"].mean()+100000, s="Country Average: {}".format(int(data["Total Cases"].mean())))
```

# States by deaths and active cases

```python
data.sort_values(by="Total Cases", ascending=False)["Death"].plot.bar(figsize=(20, 10), color="r", alpha=0.5)
data.sort_values(by="Total Cases", ascending=False)["Active"].plot.bar(figsize=(20, 10), color="g", alpha=0.5)
plt.axhline(data.Active.mean(), color="black")
```
