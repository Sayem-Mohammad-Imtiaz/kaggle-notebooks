```python
train = pd.read_csv('/kaggle/input/health-insurance-cross-sell-prediction/train.csv')
test = pd.read_csv('/kaggle/input/health-insurance-cross-sell-prediction/test.csv')
```

```python
train['Vehicle_Age'] = train['Vehicle_Age'].map({'< 1 Year':0, '1-2 Year':1, '> 2 Years':2})
test['Vehicle_Age'] = test['Vehicle_Age'].map({'< 1 Year':0, '1-2 Year':1, '> 2 Years':2})

train['Gender'] = train['Gender'].map({'Female':0, 'Male':1})
test['Gender'] = test['Gender'].map({'Female':0, 'Male':1})

train['Vehicle_Damage'] = train['Vehicle_Damage'].map({'No':0, 'Yes':1})
test['Vehicle_Damage'] = test['Vehicle_Damage'].map({'No':0, 'Yes':1})
```

```python
train_y = train['Response']
train_X = train.drop(['Response','Region_Code','id','Policy_Sales_Channel'], axis = 1)
```

```python
test_X = test.drop(['Region_Code','id','Policy_Sales_Channel'], axis = 1)
```

```python
from sklearn.linear_model import LogisticRegression


# Logistic Regression
logreg = LogisticRegression()

# Train logreg
logreg.fit(train_X, train_y)

# Predict logreg
test['Response'] = logreg.predict(test_X)
```

```python
submission = test[['id','Response']]
submission.to_csv('sample_submission.csv')
```
