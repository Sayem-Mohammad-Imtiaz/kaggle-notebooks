## Word extraction
>  Don't execute if *preproc_file* exists

```python
en_stopwords = set(nltk.corpus.stopwords.words('english'))
stemmer = nltk.PorterStemmer()
lemmatizer = nltk.WordNetLemmatizer()
```

```python
def posts_process(text: str):
    after_stem = map(stemmer.stem, nltk.word_tokenize(text))
    after_lemmatize = map(lemmatizer.lemmatize, after_stem)
    good_words = (word for word in after_lemmatize if len(word) > 2 and word not in en_stopwords and word.isalpha())
    return good_words
```

```python
def line_to_csv(p):
    return f'{p[0]};{" , ".join(p[1])}'

def to_csv(preproc_data):
    return "\n".join(map(line_to_csv , preproc_data))
```

```python
nrow = len(dataset)
with open(preproc_file, mode='w') as f:
    f.write("Type;Words\n")
for i in range(nrow):
    row = dataset.iloc[i]
    words = list(posts_process(row["posts"]))
    with open(preproc_file, mode='a') as f:
        f.write(line_to_csv((row.type, words)) + "\n")
```

```python
df_preproc = pd.read_csv(preproc_file, sep=';', header=0)
```
