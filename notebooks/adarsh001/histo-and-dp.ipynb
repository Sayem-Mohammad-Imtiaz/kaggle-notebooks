```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
airlines = pd.read_csv('../input/flight-delays/airlines.csv')
carriers = pd.read_csv('../input/flight-delays/airports.csv')
flights = pd.read_csv('../input/flight-delays/flights.csv', low_memory=False)
```

```python
airlines.head()
```

```python
carriers.head()
```

```python
flights.head()
```

```python
flights = flights.merge(airlines, how = 'left', on = 'AIRLINE')
flights.columns
```

```python
flights = flights[['ARRIVAL_DELAY' , 'AIRLINE']]
flights = flights[flights["ARRIVAL_DELAY"].between(-60,120)]
flights.head(10)
```

```python
flights["ARRIVAL_DELAY"].describe().round(3)
```

```python
from IPython.core.pylabtools import figsize
figsize(9,7)
plt.hist(flights["ARRIVAL_DELAY"] , color = "blue",edgecolor = "black",bins = int(180/5))
plt.title("Histogram of arrival Delays")
plt.xlabel("delay(mins)")
plt.ylabel("No of flights")
plt.show()
```

```python
figsize(9,7)
sns.distplot(flights["ARRIVAL_DELAY"] , color = "blue" , kde = False
            , bins = int(180/5), hist_kws = {"edgecolor": "black"}
            ,hist = True)
plt.title("Histogram of arrivals delays")
plt.xlabel("delay(mins)")
plt.ylabel("No of flights")
```

```python
figsize(10,16)
for i ,binwidth in enumerate([1,5,10,15]):
    ax = plt.subplot(2,2,i+1)
    ax.hist(flights["ARRIVAL_DELAY"],color = "Blue", edgecolor ="Black"
           , bins = int(180/binwidth))
    ax.set_xlabel( "Delay(min)" , fontsize = 12)
    ax.set_ylabel("No of flights")
    ax.set_title("histogram with binwidhts = %d"% binwidth ,fontsize = 15)
```

```python
#top five airlines with most flights
carrier_flights = pd.Series(flights.groupby("AIRLINE")["AIRLINE"].count())
carrier_flights = carrier_flights.sort_values(ascending= False)
top_five = carrier_flights.index[0:5]

```

```python
top_five_flights = flights[flights["AIRLINE"].isin(top_five)]
```

```python
figsize(10,8)
for airline in top_five:
    subset = top_five_flights[top_five_flights["AIRLINE"] == airline ]
    sns.distplot(subset["ARRIVAL_DELAY"] , kde = False , norm_hist =True
                ,hist_kws = {"edgecolor": "Black"}
                , label = airline)
    plt.title("Histogram of Delays by airlines")
    plt.xlabel("Delay(mins)")
    plt.ylabel("Normilzed Counts")
    plt.legend(prop = {"size" : 16},title = "Airlines")
```

```python
#
x1 = list(flights[flights["AIRLINE"] == "EV"]["ARRIVAL_DELAY"])
x2 = list(flights[flights["AIRLINE"] == "00"]["ARRIVAL_DELAY"])
x3 = list(flights[flights["AIRLINE"] == "AA"]["ARRIVAL_DELAY"])
x4 = list(flights[flights["AIRLINE"] == "DL"]["ARRIVAL_DELAY"])
x5 = list(flights[flights["AIRLINE"] == "WN"]["ARRIVAL_DELAY"])

colors = ['#E69F00', '#56B4E9', '#F0E442', '#009E73', '#D55E00']

names = ['WN', 'DL', 'AA', 'OO', 'EV']
```

```python
plt.hist([x1,x2,x3,x4,x5] , color = colors , label = names,
        bins = int(180/10),histtype = 'bar', density = True)
plt.legend(prop = {"size" : 15 } , title = "Airlines")
plt.xlabel("Delay(mins)")
plt.ylabel("Normalized Flights")
plt.title("Side by side histogram with multiple Airlines")
plt.show()
```

```python
figsize(10,8)
plt.hist([x1,x2,x3,x4,x5],bins= int(180/15), density = True , stacked
        = True , color = colors , label = names)
plt.legend(title = "Airlines")
plt.show()
```

```python
sns.distplot(flights["ARRIVAL_DELAY"], kde = True ,hist =  True
            ,kde_kws = {"linewidth": 3},bins = int(180/5) ,
hist_kws=dict(edgecolor="k", linewidth=2) )
plt.xlabel("delay(mins)")
plt.ylabel("Density")
plt.title("Density Plot and histogram of arrival delays")
plt.show()
```

```python
figsize(12,10)
for bandwidth in ["scott", 0.2,0.5,1,5]:
    sns.distplot(flights["ARRIVAL_DELAY"], hist = False ,kde = True,
                kde_kws = {"bw": bandwidth,"linewidth":1}, label = bandwidth)
    plt.legend(prop = {"size" : 10},title = "Bandwidth")
    plt.title("Denstiy plot with varying bandwidhts")
```

```python
for airline in ['WN', 'DL', 'AA', 'OO', 'EV']:
    subset = flights[flights["AIRLINE"] == airline ]
    sns.distplot(subset["ARRIVAL_DELAY"] , kde =  True,hist = False
                , label = airline, kde_kws = dict(shade = True))
    plt.title("Shaded density of Delays by airlines")
    plt.xlabel("Delay(mins)")
    plt.ylabel("density")
    plt.legend(prop = {"size" : 16},title = "Airlines")
```
