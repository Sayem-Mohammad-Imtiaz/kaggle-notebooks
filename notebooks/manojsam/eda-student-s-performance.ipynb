```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/kaggle/input/students-performance-in-exams/StudentsPerformance.csv")
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data["parental level of education"].value_counts()
```

```python
data["parental level of education"] = data["parental level of education"].replace({"bachelor's degree" : "Bachelor's"})
data["parental level of education"] = data["parental level of education"].replace({"master's degree" : "Master's"})
data["parental level of education"] = data["parental level of education"].replace({"some high school" : "High School"})
data["parental level of education"] = data["parental level of education"].replace({"high school" : "High School"})
data["parental level of education"] = data["parental level of education"].replace({"associate's degree" : "Associate's"})
data["parental level of education"] = data["parental level of education"].replace({"some college" : "College"})
data["parental level of education"].value_counts()
```

```python
plt.figure(figsize = (16,6))

plt.subplot(2,2,1)
sns.boxplot(x = data["gender"], y = data["math score"])
plt.subplot(2,2,2)
sns.boxplot(x = data["race/ethnicity"], y = data["math score"])
plt.subplot(2,2,3)
sns.boxplot(x = data["parental level of education"], y = data["math score"])
plt.subplot(2,2,4)
sns.boxplot(x = data["test preparation course"], y = data["math score"])

plt.show()
```

Male Gender has scored higher in Maths compared to Females

Group E has scored higher compared to other groups

Students who have finished their master's have got higher scores

Students who have completed their test preparation course have scored higher
