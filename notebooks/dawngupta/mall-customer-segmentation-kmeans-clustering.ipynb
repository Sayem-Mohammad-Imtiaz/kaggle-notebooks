```python
## import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Read the dataset

```python
df=pd.read_csv('../input/customer-segmentation-tutorial-in-python/Mall_Customers.csv')
```

```python
df.head()
```

```python
df.shape
```

#### It's a very small dataset 

#### Customer Id is just like primary key for each customer and can be removed

```python
# Remove customer ID

df.drop('CustomerID', axis=1, inplace=True)
```

```python
df.head()
```

#### Look for missing values

```python
df.isnull().sum().sort_values(ascending=False)
```

#### Change the categorical variable to numeric.

```python
# encode Female as 1 and male 0

df['Gender'] = np.where(df['Gender']=='Male',0,1)
```

```python
df.head()
```

### Now let's use elbow method to find suitable number of cluster

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(df)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 15), wcss)
plt.title('The Elbow Method')
plt.xlabel('n: Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

#### From above graph, we can observe before n=5,6 slope is decreasing sharply and after n=5,6 slope is decreasing slowly. So, we can conclude that number of cluster can be 5 or 6.

## Take n_clusters = 6 

```python
clstr = KMeans(n_clusters = 6, init = 'k-means++', random_state = 42)
```

```python
y_kmeans = clstr.fit_predict(df)

print(y_kmeans)
```

```python
df['group']=y_kmeans
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))

plt.scatter(x=df[df['group']==0]['Annual Income (k$)'],y=df[df['group']==0]['Spending Score (1-100)'], s=100, color='blue', label='C1')
plt.scatter(x=df[df['group']==1]['Annual Income (k$)'],y=df[df['group']==1]['Spending Score (1-100)'],  s=100,color='green', label='C2')
plt.scatter(x=df[df['group']==2]['Annual Income (k$)'],y=df[df['group']==2]['Spending Score (1-100)'],  s=100, color='red', label='C3')
plt.scatter(x=df[df['group']==3]['Annual Income (k$)'],y=df[df['group']==3]['Spending Score (1-100)'],  s=100,color='cyan', label='C4')
plt.scatter(x=df[df['group']==4]['Annual Income (k$)'],y=df[df['group']==4]['Spending Score (1-100)'],  s=100,color='yellow', label='C5')
plt.scatter(x=df[df['group']==5]['Annual Income (k$)'],y=df[df['group']==5]['Spending Score (1-100)'], s=100, color='orange', label='C6')



# Decorate

plt.title('Cluster of customer')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend(loc='upper left');
plt.show()
```

### From above scatter plot we can observe that C2 and C5 are getting overlap.

```python
y_kmeans = clstr.fit_predict(df)

print(y_kmeans)
```
