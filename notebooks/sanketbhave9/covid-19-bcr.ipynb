# Importing modules

```python
import pandas as pd 
import numpy as np 
from datetime import *
import bar_chart_race as bcr
```

# Importing CSV file to data frame

```python
data = pd.read_csv("../input/covid19-in-india/covid_19_india.csv")
data.head()
```

# Deleting useless columns form data frame

```python
data.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational","Cured","Deaths"],axis=1,inplace=True)
```

```python
data.head(12)
```

# Converting Date column into datetime for comparing dates

```python
dates = []
for i in list(data.Date):
    d,m,y = i.split('/')
    dates.append(date(int(y)+2000,int(m),int(d)))
data["Date"] = dates
```

```python
data.head()
```

# Deleting useless rows i.e data of Jan and Feb

```python
data.drop(data[data.Date<date(2020,3,1)].index,inplace=True)
```

```python
data.head()
```

# Convert this data into pivot table

```python
covid = data.pivot_table(index="Date",columns="State/UnionTerritory",values="Confirmed",aggfunc="sum")
```

```python
covid.head()
```

# Replacing NaN value with Zero

```python
covid = covid.fillna(0)
```

```python
covid.head()
```

# Making Bar Chart Race using this data

```python
bcr.bar_chart_race(df=covid,title="Covid 19 confirmed cases",n_bars=10,filter_column_colors=True,figsize=(5,3))
```
