```python
df=pd.read_csv('/kaggle/input/breast-cancer-wisconsin-data/data.csv')
df
```

```python
df['diagnosis']=df['diagnosis'].map({'M':1,'B':0})
df
```

```python
df=df.drop(columns=['id','Unnamed: 32'])
df
```

```python
df.isna().sum()
```

```python
X=df.drop(columns=['diagnosis'])
y=df['diagnosis']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s=StandardScaler()
```

```python
new=s.fit_transform(X)
new
```

```python
new.shape
```

```python
transformed_X=pd.DataFrame(new)
transformed_X
```

```python
transformed_X.columns=X.columns
transformed_X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(transformed_X,y,test_size=0.2)
```

```python
import sklearn
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

# Linear Regression

```python
linear=LinearRegression()
linear.fit(X_train,y_train)
```

```python
linear.score(X_test,y_test)
```

# Logistic Regression

```python
lg=LogisticRegression()
lg.fit(X_train,y_train)
```

```python
lg.score(X_test,y_test)
```

# Support Vector Classifier

```python
svc=SVC()
svc.fit(X_train,y_train)
```

```python
svc.score(X_test,y_test)
```

# Decision Tree

```python
tree=DecisionTreeClassifier(criterion='entropy')
tree.fit(X_train,y_train)
```

```python
tree.score(X_test,y_test)
```

# Random Forest

```python
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
rf.score(X_test,y_test)
```

# Based on testing through different classifiers so far, Support Vector Classifier performed the best among them with an accuracy of 96.49%
