out of 82 buildings i.e. (total entries present in dataset), 53 have the name of a male, 19 have the name of a female and 9 were named on both man and woman, whereas there was also one building which was named on two males.

We will create separate columns for male and female.
The value of a row in **male** column will be **1** for all cases except the case when the building was solely named on female in which we will assign **0** value.
The value of a row in **female** column will be **1** if the building is either solely named on single female or named on both female and male.In remaining cases, we will assign **0** value.

```python
male=[]
for i in range(len(df)):
    if df['Gender'][i] in ['M','M & F','F & M','M & M']:
        male.append(1)
    else:
        male.append(0)
```

```python
female=[]
for i in range(len(df)):
    if df['Gender'][i] in ['F','M & F','F & M']:
        female.append(1)
    else:
        female.append(0)
```

```python
df['Male']=male
df['Female']=female
df.head()
```

```python
df['Male'].sum()
```

```python
df['Female'].sum()
```

```python
plt.bar(['Male','Female'],[63,28],color=['Blue','fuchsia'])
```

```python
df_department=df.groupby('Department/Source')['Male','Female'].sum()
df_department
```

```python
df_department['Total']=df_department['Male']+df_department['Female']
df_department
```

More than **60%** (50 out of 82) of buildings in our dataset belongs to **Administrstor Department**. **19** buildings belongs to **Recreation and Parks** Department, **Library** Department ranked third in no of buildings with a figure of **11**.  

```python
plt.figure(figsize=(6,6))
plt.pie(df_department['Male'],autopct='%0.1f',labels=df_department.index)
```

```python
ind=np.arange(8)
plt.figure(figsize=(10,5))
plt.bar(ind,df_department['Male'],width=0.4,align='center',label='Male',color='Blue')
plt.bar(ind+0.4,df_department['Female'],width=0.4,align='center',label='female',color='fuchsia')
plt.title('Gender wise Distribution in Types of buildings owned')
plt.xlabel('Department')
plt.ylabel('No of buildings')
plt.xticks(ind+0.2,('AIRPORT', 'Administrator', 'LIBRARY', 'PUC', 'Port', 'REC AND PARKS',
       'RED', 'SFMTA'))
plt.xticks(rotation=90)
plt.legend(loc='best')
```

Out of the 8 types of departments/source, only two that are **Library** and **Port** showed almost **equal** distribution in male and female ownership.
In the rest 6 departments, male surpassed female in the ownership of the buildings
