```python
#Model
from sklearn.model_selection import cross_val_score, train_test_split
inputgb = ['volatile acidity', 'alcohol' , 'citric acid',  'sulphates']
target = 'quality'
#X = wine[inputgb] 
X = wine.iloc[:,:11]
Y = wine[target]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
Y_train.shape, Y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
linear_model.fit(X, Y)
print(linear_model.intercept_)
```

```python
pd.DataFrame({
  'features': X.columns.values,
  'coef': linear_model.coef_
})
```

```python
linear_model_pred = linear_model.predict(X_test).round(0)
pd.DataFrame({'prediction': linear_model_pred}).describe()
```

```python
from sklearn import metrics
linear_model_train_acc = metrics.accuracy_score(Y_train, linear_model.predict(X_train).round(0))
linear_model_test_acc = metrics.accuracy_score(Y_test, linear_model_pred)


print(f'Train Accuracy: {linear_model_train_acc*100:.2f}%')
print(f'Test Accuracy: {linear_model_test_acc*100:.2f}%')
```

```python
pd.DataFrame({
    'prediction' : linear_model_pred,
    'true value' : Y_test,
})
```

```python
# Model evaluation by MAE, MSE and RMSE
# print MAE
print("Test set evaluation")
mae = metrics.mean_absolute_error(Y_test, linear_model_pred)
print('MAE: %.3f' % mae)

# print MSE
mse = metrics.mean_squared_error(Y_test, linear_model_pred)
print('MSE: %.3f' % mse)

# print RMSE
#result of RMSE = how far the prediction from reality
rmse = np.sqrt(metrics.mean_squared_error(Y_test, linear_model_pred))
print('RMSE: %.3f' % rmse)
```

```python
linear_model_4 = LinearRegression()
linear_model_4.fit(X_train[inputgb] , Y_train)

# Predict the results
linear_model_4_pred = linear_model_4.predict(X_test[inputgb]).round(0)
pd.DataFrame({'prediction': linear_model_4_pred}).describe()
```

```python
linear_model_train_acc4 = metrics.accuracy_score(Y_train,  linear_model_4.predict(X_train[inputgb]).round(0))
linear_model_test_acc4 = metrics.accuracy_score(Y_test, linear_model_4_pred)


print(f'Train Accuracy: {linear_model_train_acc4*100:.2f}%')
print(f'Test Accuracy: {linear_model_test_acc4*100:.2f}%')
```
