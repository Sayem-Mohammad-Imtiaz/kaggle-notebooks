```python
#import libraries
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd 
import tensorflow as tf
from tensorflow.keras.utils import to_categorical 
import tensorflow.keras as K
from tensorflow.keras.layers import Dense, Dropout,Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dropout
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input
```

```python
#load data
df = pd.read_csv('/kaggle/input/sign-language-mnist/sign_mnist_train.csv')
```

```python
df.shape
```

```python
#split x and y
x_train= df.values[0:,1:]
y_train = df.values[0:,0]
```

```python
#one-hot-encode labels
y_train = to_categorical(y_train)
```

```python
#preprocessing
#reshaping train-data
x_train = x_train.reshape(-1,28,28,1)
```

```python
x_train.shape
```

```python
#example train x with its label
plt.imshow(x_train[4])
print(y_train[4])
```

```python
#preprocess (normilize data)
x_train = x_train/255
```

```python
x_train = np.stack([x_train.reshape(x_train.shape[0],28,28)]*3, axis=3).reshape(x_train.shape[0],28,28,3)
```

```python
x_train.shape
```

```python
model = Sequential()
#  Create base model of VGG16
model.add(VGG16(weights='imagenet',
                  include_top=False, pooling = 'avg',  
                  input_shape=(48, 48, 3)
                 ))
# 2nd layer as Dense 
model.add(Dense(25, activation = 'softmax'))

# Say not to train first layer (ResNet) model as it is already trained
model.layers[0].trainable = False
```

```python
model.compile(loss = K.losses.categorical_crossentropy, optimizer=K.optimizers.Adam(),
              metrics=['accuracy'])
```

```python
from tensorflow.keras.preprocessing.image import img_to_array, array_to_img
```

```python
x_train_tt = np.asarray([img_to_array(array_to_img(im, scale=True).resize((48,48))) for im in x_train])/225
```

```python
x_train_tt.shape
```

```python
model.fit(x_train_tt,y_train,epochs=5)
```

```python
test_data = pd.read_csv('/kaggle/input/sign-language-mnist/sign_mnist_test.csv')
```

```python
#split x and y in test data
x_test = test_data.iloc[:,1:]
y_test = test_data.iloc[:,0]
```

```python
x_test = x_test.to_numpy().reshape(-1,28,28,1)
x_test = x_test/255
```

```python
y_test = to_categorical(y_test)
```

```python
x_test = np.stack([x_test.reshape(x_test.shape[0],28,28)]*3, axis=3).reshape(x_test.shape[0],28,28,3)
```

```python
x_test_tt = np.asarray([img_to_array(array_to_img(im, scale=True).resize((48,48))) for im in x_test])/225
```

```python
model.evaluate(x_test_tt,y_test)
```
