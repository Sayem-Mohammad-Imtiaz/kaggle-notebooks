Verileri görselleştirirken daha genel bir bakış açısı kazanmak istersek...

```python
#pd.set_option("display.max_rows",None)
melted=pd.melt(df, id_vars="variety", var_name="features", value_name="value")
melted.iloc[:5]
```

```python
plt.figure(figsize=(10,6))
sns.violinplot(x="features", y="value", hue="variety", data=melted, inner="quart")
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="features", y="value", hue="variety", data=melted)
```

```python
plt.figure(figsize=(10,6))
sns.swarmplot(x="features", y="value", hue="variety", data=melted)
```
