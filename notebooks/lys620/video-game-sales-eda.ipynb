```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from dataprep.eda import create_report, plot
```

```python
sales = pd.read_csv('../input/videogamesales/vgsales.csv')
sales.head()
```

Rank - Ranking of overall sales

Name - The games name

Platform - Platform of the games release (i.e. PC,PS4, etc.)

Year - Year of the game's release

Genre - Genre of the game

Publisher - Publisher of the game

NA_Sales - Sales in North America (in millions)

EU_Sales - Sales in Europe (in millions)

JP_Sales - Sales in Japan (in millions)

Other_Sales - Sales in the rest of the world (in millions)

Global_Sales - Total worldwide sales.

```python
print(sales.info(),'\n')
print(sales.describe().T,'\n')
print(sales.isnull().sum(axis=0))
```

> For using 'create_report', select dtype float64 and int64

```python
sale_dtype_float = sales.loc[:, sales.dtypes == 'float64']
```

> You can check data distribution, Nan data, correlation very easy.

```python
create_report(sale_dtype_float)
```

```python
sales.head()
```

Popular genre in worldwide.

1st : Action, 2nd : Sports, 3nd : Shooter

```python
Genre = sales.groupby(sales['Genre']).sum().loc[: , ['NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
Genre['Total'] = Genre.sum(axis=1)
Genre.sort_values(by=['Total'], axis=0, ascending=False).head(10)
```

Popular publisher in worldwide.

1st : Nintendo, 2nd : Electronic Arts, 3nd : Activision

```python
Publisher = sales.groupby(sales['Publisher']).sum().loc[: , ['NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
Publisher['Total'] = Publisher.sum(axis=1)
Publisher.sort_values(by=['Total'], axis=0, ascending=False).head(10)
```

Popular platform in worldwide.

1st : PS2, 2nd : X360, 3nd : PS3

```python
Platform = sales.groupby(sales['Platform']).sum().loc[: , ['NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
Platform['Total'] = Platform.sum(axis=1)
Platform.sort_values(by=['Total'], axis=0, ascending=False).head(10)
```
