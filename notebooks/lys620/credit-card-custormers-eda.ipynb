```python
fig = make_subplots(rows=2, cols=2)
fig.add_trace(go.Bar(x=df.Gender, y=df.Gender.value_counts(), name="Gender"), row=1, col=1)
fig.add_trace(go.Bar(x=df.Education_Level, y=df.Education_Level.value_counts(), name="Edu-Level"), row=1, col=2)
fig.add_trace(go.Bar(x=df.Marital_Status, y=df.Marital_Status.value_counts(), name="Marital_Status"), row=2, col=1)
fig.add_trace(go.Bar(x=df.Card_Category, y=df.Card_Category.value_counts(), name="Card_Category"), row=2, col=2)
fig.show()
```
