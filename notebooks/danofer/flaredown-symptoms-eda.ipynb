```python
df = pd.read_csv("../input/fd-export.csv")
df.head()
```

## string keys are a waste of memory, let's use integer IDs! 

```python
df['user_id'] = pd.Categorical(df['user_id'])
df['user_id']= df.user_id.cat.codes
```

```python
df.head()
```

```python
## There are many rows where age is 0.0 / missing. replace these with NaN for consistency.
df["age"] = df.age.replace(0.0,np.nan)
```
