```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(10, 8)}); # you can change this if needed
```

Загрузите данные. Они находятся в директории input.

```python
df = pd.read_csv("../input/bigml_59c28831336c6604c800002a.csv")
df.info()
```

Каково распределение абонентов по целевому признаку **churn**? Назовите процент лояльных абонентов.

```python
churn_loyals = (sum(df['churn'] == True),sum(df['churn']==False))
plot = sns.countplot(x='churn', data=df)
# print(churn_loyals)
percent_loyals = (churn_loyals[1]/sum(churn_loyals))*100
print(percent_loyals)
```

Жители какого штата говорят по телефону больше всего? меньше всего? Обратите внимание, что нас интересуют все виды звонков.

```python
df1 = df.groupby("state")["total day calls", "total eve calls", "total night calls", "total intl calls", "customer service calls"].sum().sum(axis = 1)
print(f"Max {df1.idxmax()}: {df1[df1.idxmax()]} \nMin {df1.idxmin()}: {df1[df1.idxmin()]}")
```

Верно ли, что люди, совершающие много международных звонков, приобретают **international plan**? Не забывайте о визуализации информации.

```python
print(df.groupby("international plan")["total intl calls"].head())
sns.barplot(x="international plan",y='total intl calls', data=df)
```

Какие из числовых признаков имеют наибольшую степень корреляции? Почему?

```python
df.info()
df1 = df[["area code","number vmail messages", "total day minutes",  "total day charge", "total eve minutes","total eve calls", "total eve charge","account length", "total night minutes", "total night calls", "total night charge", "total day calls", "total intl minutes", "total intl calls", "total intl charge", "customer service calls"]]
correlation = df1.corr()
sns.heatmap(correlation)
```
