```python
%matplotlib inline
import matplotlib.pyplot as plt
# Read sets data as `sets`
sets = pd.read_csv('../input/lego-database/sets.csv')
# Create a summary of average number of parts by year: `parts_by_year`
parts_by_year = sets.groupby('year').mean
# Plot trends in average number of parts by year
fig, ax = plt.subplots()
ax.plot(parts_by_year().index, parts_by_year().num_parts)
ax.set(xlabel='Years', ylabel='Number of Parts')
ax.grid()
plt.show()
```

```python
X = np.array(parts_by_year().num_parts)
y = np.array(parts_by_year().index)
X = X.reshape(-1, 1)
y = y.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
from sklearn import ensemble
rf = ensemble.RandomForestClassifier()
rf.fit(X_train, y_train)
y_rf = rf.predict(X_test)
```

```python
print(classification_report(y_test, y_rf))
```

```python
cm = confusion_matrix(y_test, y_rf)
print(cm)
```

```python
print(accuracy_score(y_test,y_rf))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_lr = lr.predict(X_test)
print(accuracy_score(y_test,y_lr))
```

```python
probas = lr.predict_proba(X_test)
print(probas)
```

# **Wich are the LEGO's themes with the most sets ?以下都可以**

```python
# Loading sets we need
sets = pd.read_csv('../input/lego-database/sets.csv')
themes = pd.read_csv('../input/lego-database/themes.csv')
```

```python
sets.head()
```

```python
themes.head()
```

```python
#Create pandas Series with id of theme as index and count of values as value
set_theme_count = sets["theme_id"].value_counts()

#Convert it to dataframe
set_theme_count = pd.DataFrame({'id':set_theme_count.index, 'count':set_theme_count.values})
set_theme_count.head()
```

```python
set_theme_count = pd.merge(set_theme_count, themes, on='id')

set_theme_count.head()
```

```python
set_theme_count_no_parent =set_theme_count.drop(['parent_id'], axis=1)
```

```python
# Get the top 10 and plot it
set_theme_count_top_10 = set_theme_count_no_parent.sort_values(by=["count"], ascending=False)[:10]
top_10 = set_theme_count_top_10["count"]
top_10.index = set_theme_count_top_10["name"]
top_10.plot.bar()
plt.show()
```

```python
top_10.head(10)
```

```python
# Get the min 60 and plot it
set_theme_count_min_60 = set_theme_count_no_parent.sort_values(by=["count"], ascending=True)[:56]
min_60 = set_theme_count_min_60["count"]
min_60.index = set_theme_count_min_60["name"]
min_60.plot.bar()
plt.show()
```
