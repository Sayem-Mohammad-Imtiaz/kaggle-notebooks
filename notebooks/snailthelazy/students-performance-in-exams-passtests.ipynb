This Notebook is to find out who pass the exams and the passing rate.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

input_data=pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
data=input_data
data.head()
```

```python
# summary of the data
data.describe()
```

```python
# adding columns of whther a student is passed on the subject

is_math_pass=data.get(['math score'])
is_math_pass=is_math_pass.gt(50)
is_math_pass.rename(columns={'math score':'is math passed'})

is_read_pass=data.get(['reading score'])
is_read_pass=is_read_pass.gt(50)
is_read_pass.rename(columns={'reading score':'is reading passed'})

is_writing_pass=data.get(['writing score'])
is_writing_pass=is_writing_pass.gt(50)
is_writing_pass.rename(columns={'reading score':'is reading passed'})

data['is math passed']=is_math_pass
data['is reading passed']=is_read_pass
data['is writing passed']=is_writing_pass
```

```python
# extract the 3 score columns and add them together and see whether they are passed

scores_columns=['math score','reading score','writing score']
total=data.get(scores_columns).agg(['sum'],axis=1)
total.rename(columns={'sum':'total'})
data['total']=total

total_pass=data.get(['total'])
total_pass=total_pass.gt(150)
total_pass.rename(columns={'total':'is passed'})
data['is passed']=total_pass
```
