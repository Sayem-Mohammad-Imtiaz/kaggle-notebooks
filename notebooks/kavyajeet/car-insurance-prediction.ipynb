## Exploration of the dataset

first lets look at the distribution of yes and no for the vehicle insurance policy.  
lets look at it in terms of percentages:

```python
df['Response'].value_counts()/len(df)*100
```

Very few out of the given dataset have opted for the vehicle insurance. To explore what affects the response, lets explore each parameter.

In terms of percentage, around 12% of the people in the dataset has accepted the new vehicle policy

Number of acceptance between gender

```python
df['Gender'].unique()
```

```python
n_males = len(df[df['Gender']=="Male"])
n_females = len(df)-n_males
df.groupby('Gender').sum()['Response']/np.array([n_males,n_females])*100
```

around 8% of the females have responded positively to the new vehicle insurace and around 16% of male have accepted.
We can say males are slightly more likely to accept the new insurace from the data.

Now lets look at if age matters in deciding the response of the vehicle insurace.

```python
sns.distplot(df[df['Response']==1]['Age'], kde=False)
plt.show()
```

From the distribution plot of all the positive response, we can make out that people of particular age group of age 40 - 50 tend to accept the vehicle insurance. and people with age more than 60 are less likely to opt for a vehicle insurance.

Now lets look at the region code (Unique code for the region of the customer):

```python
df['Region_Code'].unique()
```

Since the values of region code is categorical, we will convert the exisiting the datatype, float to str

```python
df['Region_Code'] = df['Region_Code'].apply(lambda x: str(int(x)))
df['Region_Code'].sample(2)
```

```python
fig = plt.figure()
fig.set_size_inches(15,10)
region = df.groupby('Region_Code').sum().sort_values(by="Response",ascending = False).reset_index()
region = region[['Region_Code','Response']]
region.head(3)
```

```python
fig = plt.figure()
fig.set_size_inches(15,5)
plt.bar(region.Region_Code, region.Response)
plt.title("Response amongst different regions")
plt.xlabel("Region Code")
plt.ylabel("Number of positive Response")
plt.show()
```

We can clearly see that region code = 28 has significantly the highest number of positive reponses for the new vehicle insurace policy followed by 8 and 41. 
We can conclude that the client should priortize the vehicle insurance policy to the people belonging to this region code.
