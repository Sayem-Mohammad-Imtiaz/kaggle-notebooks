# 2. Create Model
> # 2.1 Create Machine Learning by uing 2015 data
Let's predict it by machine learning method. Firstly, create a model by using 2015 data.

```python
from sklearn.tree import DecisionTreeRegressor #DecisionTree Regressor
from sklearn.ensemble import RandomForestRegressor # RandomForest Regressor

from sklearn.model_selection import train_test_split
from sklearn import preprocessing # sklearn preprocessing, if it's necessary

# Squeeze data
# dataset2015_y = df2015['Happiness Score'] # Pandas.Series (x.values -> numpy.ndarray)
dataset2015_y = df2015[['Happiness Score']] # Pandas.DataFrame
# dataset2015_x = df2015[['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)', 'Dystopia Residual']]
# Delete 'Dystopia Residual', since this is not applicable for the year 2018 and 2019.
dataset2015_x = df2015[['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)']]

(train_x, test_x, train_y, test_y) = train_test_split(dataset2015_x, dataset2015_y, test_size=0.2, random_state=0) # just practice

# Create model
# Regression models
RandomForest_model = RandomForestRegressor()
DecisionTree_model = DecisionTreeRegressor()

# Learning model (just compare DecisionTree vs RandomForest)
DecisionTree_model.fit(train_x, train_y)
RandomForest_model.fit(train_x, train_y)
# DecisionTree_pred_y = DecisionTree_model.predict(test_x)
# RandomForest_pred_y = RandomForest_model.predict(test_x)
print(f"Decision Tree accuracy score: {DecisionTree_model.score(test_x, test_y)}") #*.socre -->> R^2. Best case is R^2=1.0
print(f"Random Forest accuracy score: {RandomForest_model.score(test_x, test_y)}")
# print(f"Accuracy:{accuracy_score(test_y, predict_y)}") # accuracy_score is applicable for classification model
```

It looks like ***random forest algorithm*** is better than ***decision tree algorithrm***.
Let's do **grid search** using Random Forest Regressor

> # 2.2 Define Best Parameters by GridSearch

```python
from sklearn.model_selection import GridSearchCV

model_param_grid = {RandomForestRegressor(): {"n_estimators": [i for i in range(100,120)],
                                                  "criterion": ["mse","mae"],
                                                  "max_depth": [j for j in range(3,11)],
                                                  "random_state":[0]}}

# initialization
best_score = 0 
best_param = None

for model, param in model_param_grid.items():
    reg = GridSearchCV(model, param)
    reg.fit(train_x, train_y)
    score = reg.score(test_x, test_y)
    if best_score < score:
        best_score = score
        best_param = reg.best_params_ # .best_param_ -->> output best parameter
        
print(f"best parameters: {best_param}")
print(f"best score: {best_score}")
# Run this code takes time..., please wait.
# Running Result:
# best parameters: {'criterion': 'mae', 'max_depth': 5, 'n_estimators': 111, 'random_state': 0}
# best score: 0.6759142902560764
```

Result:
best parameters: {'criterion': 'mae', 'max_depth': 5, 'n_estimators': 111, 'random_state': 0}
best score: 0.6759142902560764

Let's decide happiness score model as following:
1. Method: Randomforest
2. criterion: mae
3. max_depth: 5
4. n_estimators: 111

```python
# Since Dystopia Residual is not applicable for 2018 and 2019 year, let's delete it.
dataset2016_y = df2016[['Happiness Score']]
dataset2016_x = df2016[['Economy (GDP per Capita)', 'Family', 'Health (Life Expectancy)', 'Freedom', 'Trust (Government Corruption)']]

dataset2017_y = df2017[['Happiness.Score']]
dataset2017_x = df2017[['Economy..GDP.per.Capita.', 'Family', 'Health..Life.Expectancy.', 'Freedom', 'Trust..Government.Corruption.']]

dataset2018_y = df2018[['Score']]
dataset2018_x = df2018[['GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Perceptions of corruption']]

dataset2019_y = df2019[['Score']]
dataset2019_x = df2019[['GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Perceptions of corruption']]

reg = RandomForestRegressor(n_estimators=104, criterion="mse", max_depth=11, random_state=0) # set model parameters based on 2015 year data
reg.fit(train_x, train_y)
print(f"Random Forest 2015 accuracy score: {reg.score(test_x, test_y)}")
print(f"Random Forest 2016 accuracy score: {reg.score(dataset2016_x, dataset2016_y)}")
print(f"Random Forest 2017 accuracy score: {reg.score(dataset2017_x, dataset2017_y)}")
# print(f"Random Forest 2018 accuracy score: {reg.score(dataset2018_x, dataset2018_y)}") # not working because some cells are NaN
# Change from NaN value to 0 the year 2018.
dataset2018_y = dataset2018_y.fillna(0)
dataset2018_x = dataset2018_x.fillna(0)
print(f"Random Forest 2018 accuracy score: {reg.score(dataset2018_x, dataset2018_y)}")
print(f"Random Forest 2019 accuracy score: {reg.score(dataset2019_x, dataset2019_y)}")
```

> # 3.2 Apply Sample Moving Average to Smooth

```python
# SMA = Sample Moving Average
SMA3_2015_y_pred = dataset2015_y_pred.rolling(window=3).mean()
SMA3_2015_y_pred = SMA3_2015_y_pred.fillna(0)
SMA5_2015_y_pred = dataset2015_y_pred.rolling(window=5).mean()
SMA5_2015_y_pred = SMA5_2015_y_pred.fillna(0)
SMA7_2015_y_pred = dataset2015_y_pred.rolling(window=7).mean()
SMA7_2015_y_pred = SMA5_2015_y_pred.fillna(0)

fig = plt.figure(figsize=(24, 8))
ax1 = fig.add_subplot(121)
ax1.plot(dataset2015_rank, dataset2015_score, marker='o', color='black', alpha=0.6, label='original')
ax1.plot(dataset2015_rank, dataset2015_y_pred, marker='o', color='red', alpha=0.6, label='predicted')
ax1.plot(dataset2015_rank, SMA3_2015_y_pred, marker='s', color='green', alpha=0.6, label='3 SMA')
ax1.plot(dataset2015_rank, SMA5_2015_y_pred, marker='o', color='blue', alpha=0.6, label='5 SMA')
ax1.plot(dataset2015_rank, SMA7_2015_y_pred, marker='v', color='magenta', alpha=0.6, label='7 SMA')
ax1.grid(True)
ax1.legend()
ax2 = fig.add_subplot(122)
ax2.plot(dataset2015_rank, dataset2015_score, color='black', linestyle='solid', linewidth=3.0, label='original')
ax2.plot(dataset2015_rank, SMA3_2015_y_pred, marker='s', color='none', markeredgecolor='green', label='3 SMA')
ax2.plot(dataset2015_rank, SMA5_2015_y_pred, marker='o', color='none', markeredgecolor='blue', label='5 SMA')
ax2.plot(dataset2015_rank, SMA7_2015_y_pred, marker='v', color='none', markeredgecolor='magenta', label='7 SMA')
ax2.grid(True)
ax2.legend()
plt.show()
```
