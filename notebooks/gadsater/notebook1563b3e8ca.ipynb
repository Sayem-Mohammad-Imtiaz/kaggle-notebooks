```python
categories = ['Poor', 'Average', 'Good', 'Great']
elec_df['All_Schools_Category'] = elec_df['All Schools'].apply(lambda row: categories[int(abs(row - 1) // 25)])
```

```python
state_locations = {'Andaman & Nicobar Islands':'south', 
 'Andhra Pradesh':'south', 
 'Arunachal Pradesh':'north',
 'Assam':'east',
 'Bihar':'east', 
 'Chandigarh':'north', 
 'Chhattisgarh':'east',
 'Dadra & Nagar Haveli':'west', 
 'Daman & Diu':'west', 
 'Delhi':'north', 
 'Goa':'west', 
 'Gujarat':'west',
 'Haryana':'north', 
 'Himachal Pradesh':'north', 
 'Jammu And Kashmir':'north', 
 'Jharkhand':'east',
 'Karnataka':'south', 
 'Kerala':'south', 
 'Lakshadweep':'south', 
 'Madhya Pradesh':'west',
 'Maharashtra':'west', 
 'Manipur':'east', 
 'Meghalaya':'east', 
 'Mizoram':'east', 
 'Nagaland':'east',
 'Odisha':'east', 
 'Puducherry':'south',
 'Punjab':'north', 
 'Rajasthan':'north', 
 'Sikkim':'north',
 'Tamil Nadu':'south', 
 'Telangana':'south', 
 'Tripura':'east', 
 'Uttar Pradesh':'north',
 'Uttarakhand':'north', 
 'West Bengal':'east', 
 'All India':'center'}

elec_df['State_Locations'] = elec_df['State_UT'].apply(lambda row: state_locations[row])
```

```python
elec_df
```

```python
table = pd.crosstab(elec_df['State_Locations'][elec_df['State_Locations'] != 'center'], elec_df['All_Schools_Category'])
table
```

```python
from scipy.stats import chi2_contingency
 
stat, p, dof, expected = chi2_contingency(table)

alpha = 0.05

print('significance=%.3f, p=%.3f' % (alpha, p))
if p <= alpha:
    print('Variables are associated (reject H0)')
else:
    print('Variables are not associated(fail to reject H0)')
    
# This implies that whether schools recieve electricity and whether the state belongs to North, South, East or West are related.
```
