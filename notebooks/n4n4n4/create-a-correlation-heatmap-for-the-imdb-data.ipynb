Creating a correlation heatmap is a good workflow for this kind of data. As you would expect a variable like "aspect_ratio" displays little correlation with the other variables, while the film's gross seems to be correlated with the number of votes, reviews, and likes.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv("../input/movie_metadata.csv")
df.head()
```

```python
sns.heatmap(df.corr(), cmap="coolwarm")
```
