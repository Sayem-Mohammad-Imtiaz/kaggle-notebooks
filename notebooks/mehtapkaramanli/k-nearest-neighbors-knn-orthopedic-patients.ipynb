```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
patients = pd.read_csv("/kaggle/input/biomechanical-features-of-orthopedic-patients/column_2C_weka.csv")
patients.tail()
```

```python
patients.nunique()
```

```python
patients["class"].unique()
```

```python
for i in range(len(patients)):
    if(patients["class"].iloc[i]=='Normal'):
        patients["class"].iloc[i]=0
    elif(patients["class"].iloc[i]=='Abnormal'):
        patients["class"].iloc[i]=1
patients["class"]=patients["class"].astype(float)
```

```python
patients.info()
```

```python
#
y = patients['class'].values
x_data = patients.drop(["class"],axis=1)
```

```python

# normalization 
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data))
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state=1)
```

```python
# knn model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3) # n_neighbors = k
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
print(" {} nn score: {} ".format(3,knn.score(x_test,y_test)))
```

```python
# %%
# find k value
score_list = []
for each in range(1,15):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))
    
plt.plot(range(1,15),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```

```python
# %%
# find k value
score_list = []
for each in range(10,20):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))
    
plt.plot(range(10,20),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```
