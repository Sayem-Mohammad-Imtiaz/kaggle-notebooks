```python


import numpy as np 
import pandas as pd


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_data=pd.read_csv('/kaggle/input/loan-prediction-problem-dataset/train_u6lujuX_CVtuZ9i.csv')
train_data.head()
```

```python
# graphically checking Applicant income and Loan amount

import matplotlib.pyplot as plt


plt.scatter(train_data['ApplicantIncome'],train_data['LoanAmount'],s=120,edgecolor='black',linewidth=1)
plt.xlabel('ApplicantIncome')
plt.ylabel('LoanAmount')
plt.tight_layout()
```

```python
#Correlation between all features

correlation=train_data.corr()
print(correlation)
```

```python
train_data.isnull().sum()
```

```python
# Filling null values according to the most occuring values

train_data['Gender'].value_counts()
```

```python
train_data['Gender'].fillna("Male",inplace=True)
train_data['Married'].fillna("yes",inplace=True)
train_data['Dependents'].fillna(0,inplace=True)
train_data['Self_Employed'].fillna('no',inplace=True)
train_data['LoanAmount'].fillna(0,inplace=True)
train_data['Loan_Amount_Term'].fillna(0,inplace=True)
train_data['Credit_History'].fillna(0,inplace=True)
```

```python
train_data.isnull().sum()
```

```python
# Assigning dummy values to categorical data

from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()
label=le.fit_transform(train_data["Loan_ID"])
train_data['Loan_ID']=label

le_2=LabelEncoder()
label_2=le_2.fit_transform(train_data["Gender"])
train_data['Gender']=label_2

le_3=LabelEncoder()
label_3=le_3.fit_transform(train_data["Married"])
train_data['Married']=label_3

le_4=LabelEncoder()
label_4=le_4.fit_transform(train_data["Education"])
train_data['Education']=label_4

le_6 = LabelEncoder()
label_6 = le_6.fit_transform(train_data['Property_Area'])
train_data['Property_Area']=label_6

le_7 = LabelEncoder()
label_7 = le_7.fit_transform(train_data['Loan_Status'])
train_data['Loan_Status']=label_7

le_8 = LabelEncoder()
label_8 = le_8.fit_transform(train_data['Self_Employed'].astype(str))
train_data['Self_Employed']=label_8

le_9 = LabelEncoder()
label_9 = le_9.fit_transform(train_data['CoapplicantIncome'])
train_data['CoapplicantIncome']=label_9

le_10=LabelEncoder()
label_10=le_10.fit_transform(train_data['LoanAmount'])
train_data['LoanAmount']=label_10

le_11=LabelEncoder()
label_11=le_11.fit_transform(train_data['Loan_Amount_Term'])
train_data['Loan_Amount_Term']=label_11

le_12=LabelEncoder()
label_12=le_12.fit_transform(train_data[('Credit_History')])
train_data['Credit_History']=label_12

# Dependents has some 3+ values so replace it with 0

train_data['Dependents'].replace('3+',0,inplace=True)
```

```python
train_data.head()
```

```python
# Splitting data into X and Y

X=train_data[['Loan_ID','Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term',
             'Credit_History','Property_Area']]
Y=train_data['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.25)
```

```python
# Using SVM first for predictions

from sklearn.svm import SVC

model=SVC()
model.fit(X,Y)
```

```python
# Predicted rates

y_pred=model.predict(xtest)
print(y_pred)
```

```python
#Calcuating model accuracy

from sklearn.metrics import accuracy_score

clf_accuracy=accuracy_score(y_pred,ytest)
print(clf_accuracy)
```

```python
from sklearn import tree

model_1=tree.DecisionTreeClassifier()
model_1.fit(X,Y)
```

```python
y_Pred=model_1.predict(xtest)
print(y_Pred)
```

```python
# Random forest

from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier()
clf.fit(X,Y)
```

```python
# Random forest gives more better flexibility as compare to SVM

ypred=clf.predict(xtest)
print(ypred)
```

```python
from sklearn.metrics import accuracy_score

clf_accuracy=accuracy_score(ypred,ytest)
print(clf_accuracy)
```
