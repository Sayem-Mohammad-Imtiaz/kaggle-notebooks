## The most installed games 

```python
# Drop the values where k in it
y = df['installs'][df['installs'] != '100.0 k']
y = y[y != '500.0 k']
name = df['title'][df['installs'] != '100.0 k']
name = name[df['installs'] != '500.0 k']
# Replace the M with nothing
x = y.str.replace(' M', '')
# Convert to numeric
x = pd.to_numeric(x)
# Make a data frame
_df = pd.DataFrame([name,x]).T
# Get the top 10
_df.sort_values(by=['installs'], ascending=False)[:10].reset_index(drop=True)
```

## The most rattings 

```python
x = df['title']
y = df['total ratings']
# Make a data frame
_df = pd.DataFrame([x,y]).T
# Get the top 10
_df.sort_values(by=['total ratings'], ascending=False)[:10].reset_index(drop=True)
```

## Paid or Not

```python
plt.figure(figsize=(10,6))
sns.countplot(y=df['paid'], palette=sns.color_palette("husl"))
plt.title('Paid or Not', size=20)
plt.show()
```
