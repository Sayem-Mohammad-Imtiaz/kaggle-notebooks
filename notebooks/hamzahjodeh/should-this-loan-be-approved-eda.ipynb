```python
plt.figure(figsize=(10,6))
sns.countplot(x=df['RevLineCr'])
plt.title('Revolving line of credit', size=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x=df['LowDoc'])
plt.title('LowDoc Loan Program', size=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x=df['BalanceGross'])
plt.title('Gross amount outstanding', size=20)
plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['CreateJob'])
plt.title('Number of jobs created', size=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x=df['NewExist'])
plt.title('1 = Existing business, 2 = New business', size=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(pd.to_datetime(df['DisbursementDate']), color='purple')
plt.title('Disbursement date')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['BalanceGross'], color='g')
plt.title('Gross amount outstanding', size=20)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(df['CreateJob'])
plt.ylabel('Number of jobs', size=15)
plt.title('Number of jobs created', size=20)
plt.show()
```
