```python
merge_2015 = pd.merge(data_2015, data_2016)
merged_first = pd.merge(merge_2015, data_2017)
merged_second = pd.merge(data_2018, data_2019)
```

## Happiness Score - GDP Per Capita Relation / Correlation Analysis (2015-2016-2017)

```python
x = merged_first['Economy (GDP per Capita)_2015']
y = merged_first['Happiness Score_2015']

x_ = merged_first['Economy (GDP per Capita)_2016']
y_ = merged_first['Happiness Score_2016']

_x = merged_first['Economy (GDP per Capita)_2017']
_y = merged_first['Happiness Score_2017']

gdp_happiness_corr_2015 = x.corr(y)
gdp_happiness_corr_2016 = x_.corr(y_)
gdp_happiness_corr_2017 = x_.corr(_y)

print("Correlation 2015: ", gdp_happiness_corr_2015)
print("Correlation 2016: ", gdp_happiness_corr_2016)
print("Correlation 2017: ", gdp_happiness_corr_2017)

plt.plot(x, y, '.', label='2015', color='r');
plt.plot(x_, y_, '+', label='2016', color='g');
plt.plot(_x, _y, 'x', label='2017', color='b');

m, b = np.polyfit(x, y, 1)
m_, b_ = np.polyfit(x_, y_, 1)
_m, _b = np.polyfit(_x, _y, 1)

plt.plot(x, m*x + b, label='2015', color='r');
plt.plot(x_, m_*x_ + b_, label='2016', color='g');
plt.plot(_x, _m*_x + _b, label='2017', color='b');

plt.xlabel('GDP Per Capita');
plt.ylabel('Happiness Score');
plt.title("GDP Per Capita - Happines Score Correlation 2015-2016-2017");

plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.);
```
