```python
from sklearn.preprocessing import OneHotEncoder,LabelEncoder, OrdinalEncoder
from sklearn.compose import make_column_transformer
from sklearn.model_selection import train_test_split
```

```python
y=data.Churn
y.head()
```

```python
le=LabelEncoder()
y=le.fit_transform(y)
y
```

```python
nom_cols=["gender","InternetService","Contract","PaymentMethod"]
ord_cols=["Partner","Dependents","PhoneService","MultipleLines","OnlineSecurity","OnlineBackup","DeviceProtection","TechSupport","StreamingTV","StreamingMovies","PaperlessBilling"]
```

```python
trans_cols=make_column_transformer((OneHotEncoder(),nom_cols),
                                    (OrdinalEncoder(),ord_cols),
                                   remainder="passthrough")
```

```python
trans_cols.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
x.head()
```

```python
from sklearn.linear_model import LinearRegression
le=LinearRegression()
from sklearn.pipeline import make_pipeline
pipe=make_pipeline(trans_cols,le)
pipe.fit(x_train,y_train)
```

```python
pipe.fit(x_train,y_train)
pred=pipe.predict(x_test)
from sklearn.metrics import mean_squared_error
mean_squared_error(pred,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5)
from sklearn.pipeline import make_pipeline
pipe=make_pipeline(trans_cols,model)
pipe.fit(x_train,y_train)
```

```python
pipe.fit(x_train,y_train)
pred=pipe.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(pred,y_test)
```

```python
from sklearn.svm import SVC
model=SVC()
from sklearn.pipeline import make_pipeline
pipe=make_pipeline(trans_cols,model)
pipe.fit(x_train,y_train)
```

```python
pipe.fit(x_train,y_train)
pred=pipe.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(pred,y_test)
```

```python
from sklearn.svm import SVC
default=SVC(kernel="rbf")
from sklearn.pipeline import make_pipeline
pipe=make_pipeline(trans_cols,default)
pipe.fit(x_train,y_train)
```

```python
pipe.fit(x_train,y_train)
pred=pipe.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(pred,y_test)
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt
cm1=confusion_matrix(y_test,pred)

plt.figure()
df_cm=pd.DataFrame(cm1, range(2),range(2))
sns.set(font_scale=1.4)  #for label size
sns.heatmap(df_cm, annot=True,annot_kws={"size":16})
```
