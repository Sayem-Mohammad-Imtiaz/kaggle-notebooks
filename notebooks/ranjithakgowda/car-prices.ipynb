```python
x=data.drop(columns=['Selling_Price'])
y=data[['Selling_Price']]
```

```python
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=50)
```

```python
y.head()
```

```python
nom_col=[3]
ord_col=[4,5]
sc_col=[2]
```

```python
x_train
```

```python
data.select_dtypes(exclude=[int,float])
```

```python
trans_data=make_column_transformer((OneHotEncoder(sparse=False),nom_col),(OrdinalEncoder(),ord_col),(StandardScaler(),sc_col),remainder='passthrough')
```

```python
model=LinearRegression()
```

```python
set_config(display='diagram')
pipe=make_pipeline(trans_data,model)
pipe
```

```python
pipe.fit(x_train,y_train)
```
