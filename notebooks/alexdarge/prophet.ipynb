```python
df_btc=pd.read_csv('data.csv')
```

```python
df_btc=df_btc[['ds','y']]
df_btc
```

```python
from fbprophet import Prophet
#appel fonction
df_prophetb = Prophet()
#fit
df_prophetb.fit(df_btc)
```

```python
# prediction sur 1 an
fcast_time=365   
df_forecastb = df_prophetb.make_future_dataframe(periods= fcast_time, freq='D')
df_forecastb.tail(10)
```

```python
# prediction
df_forecastb = df_prophetb.predict(df_forecastb)
```

```python
df_forecastb.head()
```

```python
df_forecastb[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
#graphe
import matplotlib.pyplot as plt
df_prophetb.plot(df_forecastb, xlabel = 'Date', ylabel = "Prix du bitcoin et prédiction sur l'année prochaine")
plt.show()
```

```python
#graphe
fig = df_prophetb.plot_components(df_forecastb)
plt.show()
```
