**Handling the missing values **

Check the number of nulls/missing values in the dataset

```python
df.isna().sum()
```

Making dish_liked null values to zero

```python
df = df[df.dish_liked.isna()==False]
```

**List the name of columns**

```python
df.columns
```

Deleting the rows that are not required

```python
df = df.drop(["url","votes","rest_type","listed_in(type)"],axis=1)
```

```python
df.head()
```

**Renaming Columns**

```python
df.rename(columns={'approx_cost(for two people)':'avg_cost','listed_in(type)':'listed type'},in place=True)
```

```python
df.columns
```

```python
df.head()
```

```python
df.info()
```

**Transformations**

Converting column avg_cost to float type

```python
df['avg_cost']=df['avg_cost'].astype(str)
df['avg_cost']=df['avg_cost'].apply(lambda x:x.replace('','') )
df['avg_cost']=df[avg_cost'].astype(float)
                  df.info()
```

**Taking the list of columns with rating >5**

```python
newdf_rate=df[['name','rate']].groupby(['rate'])
newdf_rate=newdf_rate.filter(lambda x:x.mean()>5)
newdf_rate=newdf_rate.sort_values(by=['rate'])
newdf_rate
```
