```python
df = pd.read_csv('/kaggle/input/heart-attack-prediction/data.csv')
df.head()
```

We are going to find the null values presented in the datasets.

```python
df.isnull().sum()
```

We are dropping three columns, which is necessary to our dataset.

```python
x = df.drop(columns=['slope','ca','thal','fbs'])
print(x)
```

```python
y = df['fbs']
print(y)
```

Label Encoding is needed to convert the string value to integer.

```python
from sklearn import preprocessing 
label_encoder = preprocessing.LabelEncoder()  
x= x.apply(label_encoder.fit_transform)
print(x)
```

```python
y= label_encoder.fit_transform(y)
print(y)
```

Predictions using Naive-Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)
```

```python
nb = GaussianNB()
nb.fit(x_train, y_train)
```

```python
y_pred = nb.predict(x_test)
print(y_pred)

print(accuracy_score(y_test, y_pred))
```

Predictions using Randomforest

```python
from sklearn import model_selection
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
# random forest model creation
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
# predictions
rfc_predict = rfc.predict(x_test)

print("Accuracy:",accuracy_score(y_test, rfc_predict))
```
