# Проверим корреляцию признаков

```python
from seaborn import heatmap, PairGrid
import matplotlib.pyplot as plt

correlation_matrix = np.array([[np.corrcoef(X_y[i], X_y[j])[0][1] for i in range(len(X_y.T) - 1)] for j in range(len(X_y.T) - 1)])
df_corr_mat = pd.DataFrame(correlation_matrix, columns = df.columns[:-1], index = df.columns[: -1])

fig = plt.figure(figsize = (5, 4))
ax = heatmap(df_corr_mat, annot = True)
plt.xlabel('')
```

Видим очень сильную корреляцию между множеством признаков. Долготу мы оставим. Видим, что дата тоже меньше других коррелирует, оставим ее тоже. Попробуем оставить все остальыне признаки тоже. Все сильно коррелирует, и по сути все важно. 


```python
columns = list(df.columns[:-2])
y_name = df.columns[-2]
columns.insert(0, 'ones')
columns
```

```python
X, y = df[columns], df[y_name]
XT = X.T
print(f'Размеры: {X.shape, y.shape}')
print(f'Определитель = {np.linalg.det(XT @ X)}')
```

```python
B = np.linalg.inv(XT @ X) @ (XT @ y)
print(f'Коеффициенты B = {B}')
```

```python
predict = X @ B
print(f'MSE = {np.mean((predict - y) ** 2)}')
```

```python
from sklearn.linear_model import LinearRegression as LR

model = LR()
model.fit(X, y)
predict = model.predict(X)

print(f'MSE = {np.mean((predict - y) ** 2)}')
print(f'Коеффициенты LR = {model.coef_}')
```
