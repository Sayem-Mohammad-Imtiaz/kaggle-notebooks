```python
vacc_Greece = vacc_data.loc[vacc_data.country == 'Greece']
vacc_Greece
```

```python
vacc_Greece.set_index('date',inplace=True)
vacc_Greece
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(20,6))
sns.lineplot(data=vacc_Greece['daily_vaccinations_raw'][-29:-14])
plt.title('How many people are vaccinated daily? From 23/03 until 06/04',fontsize=25)
plt.xlabel('Date',fontsize=15)
plt.ylabel('Daily vaccinations', fontsize=15)
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(20,6))
sns.lineplot(data=vacc_Greece['daily_vaccinations_raw'][-15:])
plt.title('How many people are vaccinated daily? From 06/04 until 20/04',fontsize=25)
plt.xlabel('Date',fontsize=15)
plt.ylabel('Daily vaccinations', fontsize=15)
```

```python
fvacc_Greece_April = vacc_Greece.people_fully_vaccinated.tail(20)
nfvacc_Greece_April = vacc_Greece.people_vaccinated.tail(20)
```

```python
plt.figure(figsize=(25,6))
sns.lineplot(data=fvacc_Greece_April, label='Fully vaccinated')
sns.lineplot(data=nfvacc_Greece_April, label='Not fully vaccinated')
plt.title('People fully vs not fully vaccinated in April', fontsize=30)
plt.xlabel('Date', fontsize=20)
plt.ylabel('Number of people', fontsize=20)
```

```python
plt.figure(figsize=(20,6))
sns.lineplot(data=vacc_Greece['total_vaccinations'].tail(15))
plt.title('Total number of vaccinations per million. From 06/04 until 20/04', fontsize=20)
plt.xlabel('Date', fontsize=15)
plt.ylabel('Number of vaccinations', fontsize=15)
```
