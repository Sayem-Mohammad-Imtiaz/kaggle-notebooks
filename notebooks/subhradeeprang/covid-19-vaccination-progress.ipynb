# **HOW UNITED STATES IS PROGRESSING IN VACCINATION PROGRAMME**

United States has very good performance in daily vaccination and they also have the maximum number of people who are fully vaccinated. Now I have a intension to see how this country progressing day by day.Lets do this.....

```python
covid_data['date']=pd.to_datetime(covid_data['date'])
```

```python
covid_data['day']=pd.DatetimeIndex(covid_data['date']).day
```

```python
covid_data['year']=pd.DatetimeIndex(covid_data['date']).year
```

```python
covid_data['month']=pd.DatetimeIndex(covid_data['date']).month
```

```python
us_data=covid_data[covid_data['country']=='United States']
```

```python
plt.figure(figsize=(17,8))
sns.set(font_scale=1.4)
sns.barplot(x='day',y='people_fully_vaccinated',hue='month',palette='magma',data=us_data)
#plt.xticks(rotation=90)
plt.show()
```

**United States** is too much advance in vaccination programme. from the above graph it is observed that US start giving vaccination from January 14 and after that the fully vaccination count exponentially increased. On February 16,2021, the count is maximum. On 17th and 18th February the fully vaccination count is low or the vaccination is not given on those days.
