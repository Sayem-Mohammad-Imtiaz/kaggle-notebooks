# **Import Data**

```python
data = pd.read_csv('/kaggle/input/impact-of-covid19-pandemic-on-the-global-economy/transformed_data.csv')
df = pd.DataFrame(data)
df.head()
```

# Null Values Check and Fill Null Values

```python
df = df.fillna(df.mean())
print(df.isnull().sum())
```

# Convert Date Column to Date Type Data

```python
df["DATE"] = pd.to_datetime(df["DATE"], format = '%Y-%m-%d')
df.head(10)
```

# Statistical Description

```python
stat = df.describe()
stat
```

```python
print(df.info())
```

# Variable Selection

```python
variable = df[['HDI','TC', 'TD', 'STI', 'POP','GDPCAP']]
variable.head()
```

# Variable Correlation

```python
corr = variable.corr()

hm = go.Heatmap(
    z=corr.values,
    x=corr.index.values.tolist(),
    y=corr.index.values.tolist()
)


data = [hm]
layout = go.Layout(title="Variable Correlation Testing")

fig = dict(data=data, layout=layout)
iplot(fig, filename='labelled-heatmap')
```

From the heatmap we know which variable has correlated with GDP per Capita of country.

Here are the list of variable correlation value that is sorted from the highest to the lowest:
* POP (Population): 0.4588015
* TD (Total Deaths): 0.2998095
* HDI (Human Development Index): 0.2902757
* STI (Strigency Index): 0.2574667
* TC (Total Cases): 0.2555183

The sorting process based on the three most correlated variable to GDP per capita which are POP (population), TD(total deaths), and HDI(human development index)

# Sorting

Soerting based on Population variable

```python
sort = df.sort_values(by = 'POP', ascending=False)
print(sort.head())
```

Sorting based on Total Deaths variable

```python
sort = df.sort_values(by = 'TD', ascending=False)
print(sort.head())
```

Sorting based on Human Development Index Variable

```python
sort = df.sort_values(by = 'HDI', ascending=False)
print(sort.head())
```

Based on the sorting process with correlated variables, countries that will be the most affected are China, United States, and Norway.

# Aggregate and Groupby

```python
agg_func_selection = {'POP': ['mean'],'TD':['mean'],'HDI':['mean']}
coun = df.groupby(['COUNTRY']).agg(agg_func_selection).round(3)
print(coun)
```
