```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
cs = pd.read_csv('../input/pakistanintellectualcapitalcs/Pakistan Intellectual Capital - Computer Science - Ver 1.csv',encoding='latin1')
cs.set_index('S#',inplace=True)
cs.head(10)
```

```python
# Renaming columns names for convinience
cs = cs.rename(columns={
    "Teacher Name": "Teacher_name",
    "University Currently Teaching": "Current_university",
    "Department": "Department",
    "Province University Located": "Province",
    "Designation": "Designation",
    "Terminal Degree": "degree",
    "Graduated from": "Graduated_From",
    "Country": "Country",
    "Year": "Year",
    "Area of Specialization/Research Interests": "Specialization",
    "Other Information": "Other_information"
})
cs.head()  # Displaying initial five rows
```

```python
 #REPLACE DEPARTMENT NAME IN DATASET FOR DATA UNIQUENESS
cs['Department'] = cs['Department'].replace('Computer Science', 'Computer Sciences')
cs['Department'] = cs['Department'].replace('Department of Computer Systems & Software Engineering', 'Computer Science and Software Engineering')
cs['Department'] = cs['Department'].replace('Computer Science & Engineering', 'Computer Science and Engineering')
cs['Department'] = cs['Department'].replace('CS/IT', 'Computer Science & Information Technology')
cs['Department'] = cs['Department'].replace('CS & IT', 'Computer Science & Information Technology')
cs['Department'] = cs['Department'].replace('Computer Science and IT', 'Computer Science & Information Technology')
cs['Department'] = cs['Department'].replace('Computer Science & IT', 'Computer Science & Information Technology')
cs['Department'] = cs['Department'].replace('Computer Science and Information Technology', 'Computer Science & Information Technology')
cs['Department'] = cs['Department'].replace('Computing & Information Sciences', 'Computer and Information Sciences')
```

```python
cs['Specialization'].value_counts()
```

```python
Specializations = pd.DataFrame(cs['Specialization'])
Specializations['Specialization'].isnull().sum()  # Number of null values in 'specialization' column
```

```python
#Replaced specialization name
cs['Specialization'] = cs['Specialization'].replace('Software Engineering ', 'Software Engineering ')
cs['Specialization'] = cs['Specialization'].replace('SOFTWARE ENGINEERING', 'Software Engineering ')
cs['Specialization'] = cs['Specialization'].replace('Computer Networks',' Networking'  )
cs['Specialization'] = cs['Specialization'].replace('Networking',' Networking'  )
cs['Specialization'] = cs['Specialization'].replace('Computer Science ',' Computer Science '  )
cs['Specialization'] = cs['Specialization'].replace('COMPUTER SCIENCE',' Computer Science '  )
```

```python
Specializations['Specialization'] = Specializations.Specialization.str.lower()  # converting into lower case
Specializations['Specialization'] = Specializations.Specialization.str.replace('.', '') # removing all periods '.'
Specializations['Specialization'] = Specializations.Specialization.str.replace('and ', '') # removing all the 'and '
# As majority of the records in this cloumn have multiple areas of interets separated by commas ',' so splitting the records on the basis of commas ','
Specializations['Specialization'] = Specializations.Specialization.str.split(',')
```

```python
cs.Specialization
```

```python
Specializations.dropna(inplace=True) # Dropping the rows with null values
```

```python
Specialization="Specialization"
cs[Specialization].value_counts().sort_values()[::-1][:12].plot(kind="bar")
```

```python
cs['degree'] = cs['degree'].str.strip()\
                .replace(regex='(BS|BSCS)'           ,value='BS')\
                .replace(regex='(MCS|MSCS|M.S)'      ,value='MS')\
                .replace(regex='(MCS|MSCS|M.S|M.E)'  ,value='MS')\
                .replace(regex='(M.Phil|Mphil|MPhil)',value='MPhil')\
                .replace(regex='(Phd|Ph.D)'          ,value='PhD')\
                .replace('PhD(Scholar)','PhD').replace('PhD (Scholar)','PhD')
                            

cs['degree'].value_counts()

```
