```python
data= pd.read_csv("../input/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv")
data.head(10)
```

```python
totalCasesSorted= data.sort_values(by="Total Cases",ascending=False)
px.bar(totalCasesSorted, x="State/UTs", y="Total Cases", title="Total Covid Caess",
            labels={"value": "count", "variable": "Total Cases"})
```

```python
totalTop10Death = data.sort_values(by="Deaths",ascending=False).head(10)
totalTop10DeathFig = px.pie(totalTop10Death, values='Deaths', names='State/UTs')
totalTop10DeathFig.show()
```

```python
topBadStates = data.sort_values(by=["Active Ratio (%)","Death Ratio (%)"],ascending=False).head(10)
print(topBadStates)
topBadStatesFig = px.bar(topBadStates, x="State/UTs", y=["Active Ratio (%)", "Death Ratio (%)"], title="Wide-Form Input")
topBadStatesFig.show()
```
