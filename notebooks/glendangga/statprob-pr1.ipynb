```python
data = pd.read_csv("/kaggle/input/500-person-gender-height-weight-bodymassindex/500_Person_Gender_Height_Weight_Index.csv")
print(data)
```

```python
berat = data.Weight
berat.mean()
```

```python
berat.median()
```

```python
#modus
berat.mode()
```

```python
range = berat.max() - berat.min() + 1
print(range)
```

```python
#variance
berat.var()
```

```python
#standard deviation
berat.std()
```

```python
#percentile-15
berat.quantile(0.15)
```

```python
#percentile-90
berat.quantile(0.9)
```

```python
#interquantile range
iqr = berat.quantile(0.75) - berat.quantile(0.25)
print(iqr)
```

```python
#reasonable lower boundary
rlb = berat.quantile(0.25) - 1.5*iqr
print(rlb)
```

```python
#reasonable upper boundary
rub = berat.quantile(0.75) + 1.5*iqr
print(rub)
```

```python
#outlier
data.loc[berat<rlb]
```

```python
#plot
plt.boxplot(berat)
plt.show()
```
