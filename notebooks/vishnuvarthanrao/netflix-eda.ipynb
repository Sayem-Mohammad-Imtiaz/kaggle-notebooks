```python
unique_ratings = india_df['rating'].unique()
```

```python
unique_ratings
```

```python
ratings_data=[]
for rating in unique_ratings:
    data={
        "Rating":rating,
        "Count":(india_df['rating']==rating).sum()
    }
    ratings_data.append(data)
```

```python
ratings_data
```

```python
ratings_df =pd.DataFrame(ratings_data)
```

```python
ratings_df
```

## 2018 Data of Movies in Netflix

```python
df_2018 = india_df[india_df['release_year'] == 2018]
```

```python
df_2018
```

```python
duplicate_list=df_2018['listed_in'].unique()
```

```python
duplicate_list
```

```python
cat_list=[]
for item in duplicate_list:
    item_array =item.split(',')
    for i in item_array:
        if not i.strip() in cat_list:
            cat_list.append(i.strip())
```

```python
cat_list
```

```python
data =dict()
for item in cat_list:
    data[item]= (india_df['listed_in'].str.contains(item)).sum()


    
    
    
```

```python
keys,val = list(data.keys()),list(data.values())
x = np.arange(len(keys))
plt.bar(x,val)
plt.xticks(x,keys,rotation=90,fontSize=12)
plt.title("Distribution of Genres ",fontSize=20)
plt.show()
```
