Lets load out Dataset

```python
df = pd.read_csv("../input/diabetes.csv")
```

```python
df.info()
```

So, 768 entries seems that we don't have much data. The good thing is that we don't have any missing values.

Let's do some data visualization and plot histogram

```python
df.hist(figsize=(10,8))
```

Seems that DiabetesPedigreeFunction is nice feature to start with

```python
X = df[['DiabetesPedigreeFunction']]
y = df['Outcome']
```

We have less than 100k entries so we'll start with LinearSVC

```python
model = LinearSVC()
model.fit(X,y)
predicted = model.predict(X)
accuracy_score(y,predicted)
```

That seems to have average accuracy. Lets see if we can get better.

```python
model = KNeighborsClassifier()
model.fit(X,y)
predicted = model.predict(X)
print(accuracy_score(y,predicted))
```

```python
model = NearestCentroid()
model.fit(X, y)
predicted = model.predict(X)
accuracy_score(y,predicted)
```

KNeighbors Classifiers didn't perform up to the mark. Lets move to SVC.

```python
model = svm.SVC()
model.fit(df[['DiabetesPedigreeFunction']],df['Outcome'])
pridected = model.predict(df[['DiabetesPedigreeFunction']])
print(accuracy_score(df['Outcome'],predicted))
```

Moving towards Ensemble Classifiers.

```python
model = RandomForestClassifier()
model.fit(X,y)
predicted = model.predict(X)
accuracy_score(y,predicted)
```

That's seem to be pretty good score.

```python
model = DecisionTreeClassifier()
model.fit(X, y)
predicted = model.predict(X)
accuracy_score(y,predicted)
```

```python
model = ExtraTreesClassifier()
model.fit(X, y)
predicted = model.predict(X)
accuracy_score(y,predicted)
```

With ExtraTrees we got accurany of about 87%. Pretty nice. But let's check ahead are we overfitting our model.

```python
model = AdaBoostClassifier()
model.fit(X, y)
predicted = model.predict(X)
accuracy_score(y,predicted)
```

```python
model = GradientBoostingClassifier()
model.fit(X, y)
predicted = model.predict(X)
accuracy_score(y,predicted)
```

Dividing dataset into train data and test data

```python
from sklearn.model_selection import train_test_split
outcome=df['Outcome']
data=df[df.columns[:8]]
train,test=train_test_split(df,test_size=0.25,random_state=0,stratify=df['Outcome'])# stratify the outcome
train_X=train[train.columns[:8]]
test_X=test[test.columns[:8]]
train_Y=train['Outcome']
test_Y=test['Outcome']
```

```python
model = ExtraTreesClassifier(n_estimators=10)
model.fit(train_X, train_Y)
predicted = model.predict(test_X)
accuracy_score(test_Y,predicted)
```

```python
model = DecisionTreeClassifier()
model.fit(train_X,train_Y)
predicted = model.predict(test_X)
accuracy_score(test_Y,predicted)
```

It seems that we were overfitting our model. By diving data in test and train we got around 80% accuracy with ExtraTreesClassfier. Nice!
