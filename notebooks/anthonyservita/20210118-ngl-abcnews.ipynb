```python
import pandas as pd
import random
import markovify

#import code from data million-headlines
datasets = pd.read_csv('../input/million-headlines/abcnews-date-text.csv')
datasets.head(10)
```

```python
datasets.headline_text[random.sample(range(100), 10)]
```

```python
# size of news
model = markovify.NewlineText(datasets.headline_text,
                             state_size = 2)
# generator news
for i in range(10):
    print(model.make_sentence())
    pass

```

```python
model1 = markovify.NewlineText(datasets.headline_text,
                              state_size = 3)
model2 = markovify.NewlineText(datasets.headline_text,
                              state_size = 4)
```

```python
# size
for i in range(5):
    print(model1.make_sentence())
    pass
```

```python
for i in range(10):
    temp = model2.make_sentence()
    if temp is not None:
        print(temp)
        pass
    pass
model11 = markovify.NewlineText(datasets.headline_text,
                               state_size = 2)
model12 = markovify.NewlineText(datasets.headline_text,
                               state_size = 2)
model_combo = markovify.combine([model11, model12], [1.5, 1])

for i in range(5):
    print(model_combo.make_sentence())
    
```
