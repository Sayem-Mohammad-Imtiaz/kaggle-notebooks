## We can proceed with df_processed as there in not null data

```python
df_processed.dtypes
```

```python
df_processed['Date'] = pd.to_datetime(df_processed['Date'],format='%Y-%m')
```

```python
df_processed['Index'].replace(to_replace=["000001.SS"],value="Shangai1_China",inplace=True)
df_processed['Index'].replace(to_replace=["399001.SZ"],value="Shangai2_China",inplace=True)
df_processed['Index'].replace(to_replace=["J203.JO"],value="SouthAfrica",inplace=True)
df_processed['Index'].replace(to_replace=["GSPTSE"],value='Canada',inplace=True)
df_processed['Index'].replace(to_replace=["HSI"],value='Hongkong_China',inplace=True)
df_processed['Index'].replace(to_replace=["GDAXI"],value='German_Frankfurt',inplace=True)
df_processed['Index'].replace(to_replace=["SSMI"],value='Swiss',inplace=True)
df_processed['Index'].replace(to_replace=["IXIC"],value='Nasdaq_US',inplace=True)
df_processed['Index'].replace(to_replace=["TWII"],value='Taiwan_China',inplace=True)
df_processed['Index'].replace(to_replace=["NSEI"],value='NSE_India',inplace=True)
df_processed['Index'].replace(to_replace=["N100"],value='Nasdaq_US',inplace=True)
df_processed['Index'].replace(to_replace=["NYA"],value='NewYork_US',inplace=True)
df_processed['Index'].replace(to_replace=["N225"],value='Nikkei_Tokyo',inplace=True)

df_processed.head()
```

```python
df_processed = df_processed.drop(['Volume','Open','High','Low','Close','Adj Close'],axis=1)
df_processed
```

## Plotting the time series plots of different stock exchanges

```python
df_processed = df_processed.set_index('Date')
```

```python
df_processed.tail()
```

```python
df_processed.groupby('Index')['CloseUSD'].plot(legend=True,figsize=(15,6))
```
