**Import libraries**

```python
import pandas as pd
import numpy as np
import matplotlib as plt
```

**Import dataset**

```python
dataset = pd.read_csv('../input/netflix-shows/netflix_titles.csv')
```

```python
print(dataset.shape)
```

```python
print(dataset.head)
```

**Check missing values**

```python
#missing values 
dataset_with_na = [features for features in dataset.columns if dataset[features].isnull().sum()>1]
for feature in dataset_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4),  ' % missing values')
```

**Make a dataset copy**

```python
dataset_modified = dataset.copy()
dataset_modified
```

**Exploratory Data Analysis**

> Count distinct and group by on "type"

```python
#count distinct and group by "type"
type = dataset_modified.type.value_counts().tolist()
print(type)
```

Creat the bar chart for Movie and TV Show

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
import seaborn as sns  
```

```python

type_mt = dataset_modified['type'].value_counts()
type_mt = type_mt.sort_index(ascending=True)

type_mt.plot(kind="bar")
plt.title("Number of Movie and Tv Show on Netflix")
```

> > * Number of movies are greater than tv shows on Netflix.
