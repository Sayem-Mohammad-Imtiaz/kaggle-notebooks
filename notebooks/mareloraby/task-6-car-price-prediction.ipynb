```python
X_train, X_test, y_train, y_test = train_test_split( cars.drop('selling_price', axis=1), cars['selling_price'], test_size=0.2, random_state=42)
```

```python
y_train = np.array(y_train).reshape(-1,1)
y_test = np.array(y_test).reshape(-1,1)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
X_train= sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)

y_train= sc.fit_transform(y_train)
y_test = sc.fit_transform(y_test)
```

```python
X_train.shape
```

```python
model = keras.Sequential(
    [
        Dense(15, name="input", input_shape=(14,), activation = 'relu' ),
        Dense(30, name="hidden1", input_shape=(14,), activation = 'relu' ),
        Dense(30, name="hidden2", input_shape=(14,), activation = 'relu' ),
        Dense(1, name="output", activation = 'linear')
    ]
)
model.summary()
```

```python
model.compile(loss='mean_squared_error',optimizer='Adam')
```

```python
h = model.fit(X_train, y_train, epochs=50)
```
