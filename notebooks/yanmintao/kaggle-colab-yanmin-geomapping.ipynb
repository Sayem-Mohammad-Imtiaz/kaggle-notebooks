```python
city = 'Wuhan'
city_time_series = city_temperature[city_temperature.City == city]
city_time_series2 = city_temperature[city_temperature.City == city]
city_time_series2  
# dropping null value columns to avoid errors 
city_time_series2.dropna() 
  
```

```python
# Add year column and convert to int
city_time_series2["year"] = city_time_series2["dt"].str[0:4].astype(int)
```

```python
source = city_time_series2
source
```

```python
# Line chart to show the yearly average temperature trend in a city: Wuhan or Chongqing, a very hot city in China
alt.Chart(source).mark_line().encode(
    x='year:Q',
    # y='average(AverageTemperature):Q',
    y=alt.Y('average(AverageTemperature):Q', scale=alt.Scale(domain=[15, 20]), title = 'Yearly Average Temperature'),
    color='City:N',
    strokeDash='City'
).properties(
     width = 1000,
     height = 500
)
```

```python
chart = alt.Chart(source).mark_point().encode(
    x=alt.X('year:Q',scale=alt.Scale(domain=[1840, 2020]), title = 'Year'),
    #y='average(AverageTemperature):Q',
    #y=alt.Y('average(AverageTemperature):Q', scale=alt.Scale(domain=[15, 20]), title = 'Temperature Trend in Deg C'),
    y=alt.Y('average(AverageTemperature):Q', scale=alt.Scale(zero=False), title = 'Temperature Trend in Deg C'),
    
    color='City:N',
    strokeDash='City'
).properties(
     width = 800,
     height = 500
)
chart
```

```python
# How to print out coef, and rSquared from the linear transformation
# https://stackoverflow.com/questions/60237871/altair-access-rsquared-value-in-a-regression
# alt.Chart customization
    
line = chart.transform_regression('year','AverageTemperature').mark_line()

params = alt.Chart(source).transform_regression(
    'year', 'AverageTemperature', params=True
).mark_text(align='left').encode(
    x=alt.value(20),  # pixels from left
    y=alt.value(20),  # pixels from top
    #text='rSquared:N', 
    text= 'coef:N', # Looks like this includes the coef and R2
    opacity=alt.value(0.8),
    color=alt.value('black'),
)

chart + line + params
```
