```python
import numpy as np
import pandas as pd
```

```python
d = pd.read_csv('../input/social-network-ads/Social_Network_Ads.csv')
d.head()
```

```python
d['Gender'].replace({'Male':0, 'Female':1}, inplace = True)
```

```python
d.head()
```

```python
X = d.iloc[:,1:4]
y = d.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
X = sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2, random_state = 1)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
params = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [1,2,3,4,5,6,7,None]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(clf,param_grid = params, cv = 10, n_jobs = -1)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_estimator_
```

```python
grid.best_score_
```

```python
 grid.best_params_
```

```python
A = d.iloc[:,2:4].values
A.shape
```

```python
A_train.shape
```

```python
A_test.shape
```

```python
s = StandardScaler()
```

```python
A_train = s.fit_transform(A_train)
A_train[0:5]
```

```python
A_test = s.fit_transform(A_test)
A_test[0:5]
```

```python
# finding k METHOD 1
np.sqrt(A_train.shape[0])
```

```python
k = 17
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=k)
```

```python
knn.fit(A_train, b_train)
```

```python
b_pred = knn.predict(A_test)
```

```python
accuracy_score(b_test,b_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(b_test,b_pred)
```

```python
# METHOD 2
accuracy = []
for i in range(1,26):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(A_train,b_train)
    accuracy.append(accuracy_score(b_test, knn.predict(A_test)))
```

```python
accuracy
```
