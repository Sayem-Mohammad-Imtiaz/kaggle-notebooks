```python
data = pd.read_csv("../input/credit-card-customers/BankChurners.csv")
```

```python
data.drop(['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1', 'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'], axis = 1, inplace=True)
```

```python
data.head()
```

```python
import seaborn as sns
```

```python
sns.heatmap(data.isnull(), yticklabels=False)
```

```python
data.info()
```

```python
sex = pd.get_dummies(data['Gender'], drop_first=True)
education = pd.get_dummies(data['Education_Level'], drop_first=True)
marriage = pd.get_dummies(data['Marital_Status'], drop_first=True)
income = pd.get_dummies(data['Income_Category'], drop_first=True)
card = pd.get_dummies(data['Card_Category'], drop_first=True)
```

```python
data.drop(['Gender', 'Education_Level','Marital_Status', 'Income_Category','Card_Category'], axis = 1, inplace = True)
```

```python
data.head()
```

```python
data = pd.concat([data,sex,education,marriage,income,card], axis = 1)
```

```python
data.info()
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
stayed = pd.get_dummies(data['Attrition_Flag'], drop_first=True)
```

```python
data.drop('Attrition_Flag', axis = 1, inplace = True)
data = pd.concat([data,stayed], axis = 1)
```

```python
data.info()
```

```python
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data.drop('Existing Customer', axis = 1), data['Existing Customer'], test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
dtree_prediction = dtree.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train,y_train)
```

```python
rfc_prediction = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print("Decision Tree Results\n")
print(classification_report(y_test,dtree_prediction))
print(confusion_matrix(y_test,dtree_prediction))
```

```python
print('Random Tree Results\n')
print(classification_report(y_test,rfc_prediction))
print(confusion_matrix(y_test,rfc_prediction))
```
