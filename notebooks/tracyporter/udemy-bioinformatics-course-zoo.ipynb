Problem statement

This dataset consists of 101 animals from a zoo.
There are 16 variables with various traits to describe the animals.
The 7 Class Types are: Mammal, Bird, Reptile, Fish, Amphibian, Bug and Invertebrate

The purpose for this dataset is to be able to predict the classification of the animals, based upon the variables.
It is the perfect dataset for those who are new to learning Machine Learning.

Analyse type

```python
label = df['class_type']

target = df['class_type'].value_counts().sort_index()
target
```

```python
target.plot(kind="bar");
```

Drop animal name and target from df

```python
df = df.drop(['animal_name', 'class_type'], axis=1)
df
```

Check for missing values

```python
df.isnull().sum()
```

Create heatmap

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```

Normalise

```python
df = (df - df.min()) / (df.max() - df.min())
df
```

Standardise

```python
#df = (df - np.mean(df)) / np.std(df)
#df
```

Define X and y

```python
y = label
X = df
y.shape, X.shape
```

Split

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split( X, y, test_size=0.1, shuffle=True, random_state=1, stratify=y)
X_train.shape, y_train.shape, X_val.shape, y_val.shape
```

Select model

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

#LR = LogisticRegression()

#parameters = {'penalty':('l1', 'l2', 'elasticnet', 'none'), 'C': [0.1, 1, 10, 100], 'solver': ('newton-cg', 'lbfgs','liblinear', 'sag', 'saga'), 'max_iter':[100, 500, 1000, 10000], 'multi_class': ('auto', 'ovr', 'multinomial')}

#clf = GridSearchCV(LR, parameters).fit(X_train, y_train)
#print(clf.best_params_)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(C=10, max_iter=100, multi_class='multinomial', penalty='l2', solver='lbfgs', random_state=1).fit(X_train, y_train)

print(model.score(X_train, y_train))
```

Predict on validation set

```python
y_pred = model.predict(X_val)
print(model.score(X_val, y_val))
```

```python
#Create a confusion matrix
from sklearn.metrics import confusion_matrix

confusion_matrix(y_val, y_pred)
```
