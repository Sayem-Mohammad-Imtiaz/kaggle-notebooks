Drop id, price and date

```python
target = df['price']

df.drop(['id', 'date', 'price'], axis=1, inplace=True)
df
```

Heatmap

```python
corrmat = df.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True);
```

Drop low correlation columns

```python
df.drop(['yr_built', 'yr_renovated', 'zipcode', 'lat', 'long', 'sqft_living15', 'sqft_lot15'], axis=1, inplace=True)
df
```

Define X and y

```python
y = target
X = df
```

Standardise

```python
X = (df - np.average(X)) / (np.std(X))
X
```

Split

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.1, random_state=1)
X_train.shape, X_val.shape, y_train.shape,y_val.shape
```

Select Model

```python
from sklearn.model_selection import GridSearchCV
from sklearn import linear_model

BR = linear_model.BayesianRidge()

parameters = {'n_iter':[300, 500, 1000, 5000]}

reg = GridSearchCV(BR, parameters).fit(X_train, y_train)
print(reg.best_params_)
```

```python
from sklearn import linear_model

model = linear_model.BayesianRidge(n_iter=300).fit(X_train, y_train)
print(model.score(X_train, y_train))
```

Predict on validation set

```python
y_pred = model.predict(X_val)
print(model.score(X_val, y_val))
```

```python
from sklearn.metrics import mean_squared_error

rmse = mean_squared_error(y_val, y_pred, squared=False)
rmse
```

```python
df=pd.DataFrame({'Actual': y_val, 'Predicted':y_pred})
df
```
