Create INCOME column

```python
df['INCOME'] = 0
```

```python
income = []

for i in range(len(df)):
    if df.NAME_INCOME_TYPE[i] == "Pensioner" or df.NAME_INCOME_TYPE[i] == "Student":
        value = 0
    else:
        value = 1
    income.append(value)

df.INCOME = income
df
```

Create property column

```python
df['PROPERTY'] = 0
```

```python
property = []

for i in range(len(df)):
    if df.FLAG_OWN_CAR[i] == "Y" or df.FLAG_OWN_REALTY[i] == "Y":
        value = 1
    else:
        value = 0
    property.append(value)

df.PROPERTY = property
df
```

Create label

```python
df['LABEL'] = 0
```

```python
label = []

for i in range(len(df)):
    if df.STATUS[i] < 1 and df.INCOME[i] == 1 and df.PROPERTY[i] == 1:
        value = 1
    else:
        value = 0
    label.append(value)

df.LABEL = label
df
```

Analyse label

```python
label_good = df.LABEL.sum()
label_good
```

```python
label_good_percent = label_good / len(df)
label_good_percent
```
