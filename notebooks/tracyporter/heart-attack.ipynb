Drop target

```python
target = heart.output
heart.drop('output', axis=1,inplace=True)
heart
```

Normalise

```python
heart = (heart.max() - heart / heart.max() - heart.min())
heart
```

Define

```python
y = target
X = heart
```

Split

```python

from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split( X, y, test_size=0.1, shuffle=True, random_state=1, stratify=y)
X_train.shape, y_train.shape, X_val.shape, y_val.shape
```

Model

```python
from sklearn.svm import SVC

model = SVC(random_state=1, class_weight='balanced', kernel='linear', C=10).fit(X_train, y_train)
print(model.score(X_train, y_train))
```

Predict on val;idation set

```python
y_pred = model.predict(X_val)
print(model.score(X_val, y_val))
```

```python
from sklearn.metrics import confusion_matrix

confuse = confusion_matrix(y_val, y_pred)
confuse
```

Plot errors

```python

from sklearn.decomposition import PCA

pca_val = PCA(n_components=2)
principalComponents_val = pca_val.fit_transform(X_val)
plt.figure(figsize = (12, 8))
plt.scatter(principalComponents_val[:,0], principalComponents_val[:,1], c = y_pred == y_val - 1, alpha = .8, s = 50)
```
