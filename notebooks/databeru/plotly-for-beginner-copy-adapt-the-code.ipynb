# 1. Scatterplot<a class="anchor" id="1"></a>
## 1.1. Basic Scatterplot<a class="anchor" id="2"></a>

```python
import plotly.graph_objs as go
import numpy as np

# Generate random data
np.random.seed(0)
x = np.random.randint(1,101,100)
y = np.random.randint(1,101,100)

data = [go.Scatter(x=x, y=y, mode = 'markers')]

fig = go.Figure(data=data)
fig.show()
```

# 1.2. Scatterplot with Axis, titles and style<a class="anchor" id="3"></a>

```python
import plotly.graph_objs as go
import numpy as np
# Generate random data
np.random.seed(0)
x = np.random.randint(1,101,100)
y = np.random.randint(1,101,100)

data = [go.Scatter(x=x, 
                   y=y, 
                   mode = 'markers',
                   marker=dict(
                               size=12,
                               color='rgb(51,204,153)',
                               symbol='pentagon',
                               line = {'width':2}))]
# Add a title and a x-axis titel
layout = go.Layout(title='Title of the chart',
                  xaxis={'title':'MY X AXIS'},
                  yaxis={'title':'MY Y AXIS'},
                  hovermode='closest')

fig = go.Figure(data=data,layout=layout)
fig.show()
```

## 2.4. List comprehension for multiple values on the same chart<a class="anchor" id="9"></a>

```python
import plotly.graph_objs as go
import numpy as np
import pandas as pd

# Generate random data
np.random.seed(56)
x = np.linspace(0,1,100)
y1 = np.random.randn(100)
y2 = 2 * y1 + 7
y3 = 1.3 * y1 + 12

lst = []
for i in [x,y1,y2,y3]:
    lst.append(pd.Series(i))
dfp = pd.concat(lst, axis = 1)
dfp.columns = ['x','y1','y2','y3']

data = [go.Scatter(x=dfp['x'],y=dfp[c],
                  mode='markers+lines',
                  name=c)
       for c in dfp.columns[1:]]

layout = go.Layout(title='Line Charts')

fig = go.Figure(data=data,layout=layout)
fig.show()
```

# 3. Bar chart<a class="anchor" id="10"></a>

Use the countries from the olympic games medal dataset with the most medals.

- Import and clean the dataset
- Keep only the top 5 countries

```python
# Use the countries from the olympic games medal dataset
# with the most medals

# Import and clean the dataset
# Keep only the top 5 countries
olympic = pd.read_csv("../input/olympic-games-medal-datasetfrom-1896-to-2018/Olympic Games medal Dataset(from 1896 to 2018).csv")
olympic['Team'] = olympic['Team(IOC Code)'].apply(lambda x: x.split('(')[0])
olympic.index = olympic['Team']
olympic = olympic[['s_gold','s_silver','s_bronze','summer_total']]
olympic.columns = ['gold','silver','bronze','total']
olympic['gold'] = olympic['gold'].apply(lambda x: x.replace(',',''))
olympic['total'] = olympic['total'].apply(lambda x: x.replace(',',''))
olympic = olympic.astype({'gold': 'int64'})
olympic = olympic.astype({'total': 'int64'})
olympic.index.name = None
olympic.index = [i.replace(' ','_') for i in olympic.index]
olympic.sort_values(by='total', ascending = False, inplace = True)
olympic = olympic.iloc[:5]
```

# 4. Bubble Plots<a class="anchor" id="15"></a>

```python
import plotly.graph_objs as go
import pandas as pd

df = pd.read_csv('../input/autompg-dataset/auto-mpg.csv')

# Clean the dataset
df = df[df.horsepower != '?']
df = df.astype({'horsepower': 'int64'})

df.head()
```

## 4.1. Bubble Plots - Change the size<a class="anchor" id="16"></a>

```python
data = [go.Scatter(x=df['horsepower'],
                   y=df['mpg'],
                   mode = 'markers',
                   # The size changes depending of the number of cylinders
                   marker=dict(size=2*df['cylinders']))] 
fig = go.Figure(data = data)
fig.show()
```

## 4.2. Bubble Plots - Change the color<a class="anchor" id="17"></a>

```python
data = [go.Scatter(x=df['horsepower'],
                   y=df['mpg'],
                   mode = 'markers',
                   # The color changes depending of the number of cylinders
                   marker=dict(color=df['cylinders']))] 
fig = go.Figure(data = data)
fig.show()
```

## 4.3. Bubble Plots - Add a scale<a class="anchor" id="18"></a>

```python
data = [go.Scatter(x=df['horsepower'],
                   y=df['mpg'],
                   mode = 'markers',
                   marker=dict(color=df['cylinders'],
                                showscale=True) # Add a scale
                  )] 
fig = go.Figure(data = data)
fig.show()
```

## 4.4. Bubble Plots - Add a title for mouse over<a class="anchor" id="19"></a>

```python
data = [go.Scatter(x=df['horsepower'],
                   y=df['mpg'],
                   text=df['car name'],
                   mode = 'markers',
                   marker=dict(color=df['cylinders'],
                                showscale=True) # Add a scale
                  )] 
fig = go.Figure(data = data)
fig.show()
```

## 4.5. Bubble Plots - Change size and color<a class="anchor" id="20"></a>

```python
data = [go.Scatter(x=df['horsepower'],
                   y=df['mpg'],
                   text=df['car name'],
                   mode = 'markers',
                   marker=dict(color=df['cylinders'],
                               size=2*df['cylinders'],
                               showscale=True) # Add a scale
                  )] 
layout = go.Layout(title='Bubble Plots - Change size and color')
fig = go.Figure(data = data, layout = layout)
fig.show()
```

# 5. Boxplot<a class="anchor" id="22"></a>
## 5.1. Basic Boxplot<a class="anchor" id="23"></a>

```python
from plotly import graph_objects as go
y = [1,14,14,15,19,20,20,20,25,28,28,29,35,50]

data = [go.Box(y=y)]
fig = go.Figure(data=data)
fig.show()
```

## 5.2. Boxplot & visibles points<a class="anchor" id="24"></a>
### 5.2.1. Display the points on the left side



**boxpoints:**
    If "outliers", only the sample points lying outside the
    whiskers are shown If "suspectedoutliers", the outlier
    points are shown and points either less than 4*Q1-3*Q3
    or greater than 4*Q3-3*Q1 are highlighted (see
    `outliercolor`) If "all", all sample points are shown
    If False, only the box(es) are shown with no sample
    points Defaults to "suspectedoutliers" when
    `marker.outliercolor` or `marker.line.outliercolor` is
    set. Defaults to "all" under the q1/median/q3
    signature. Otherwise defaults to "outliers".

```python
from plotly import graph_objects as go
y = [1,14,14,15,19,20,20,20,25,28,28,29,35,50]

data = [go.Box(y=y,
              boxpoints='all')]
fig = go.Figure(data=data)
fig.show()
```

### 5.2.2. Change jitter to adapt the alignment of the points

**jitter:**
    Sets the amount of jitter in the sample points drawn.
    If 0, the sample points align along the distribution
    axis. If 1, the sample points are drawn in a random
    jitter of width equal to the width of the box(es).

```python
from plotly import graph_objects as go
y = [1,14,14,15,19,20,20,20,25,28,28,29,35,50]

data = [go.Box(y=y,
              boxpoints='all',
              jitter=0.1)]
fig = go.Figure(data=data)
fig.show()
```

### 5.2.3. Change the position where the points are displayed 

**pointpos:**
    Sets the position of the sample points in relation to
    the box(es). If 0, the sample points are places over
    the center of the box(es). Positive (negative) values
    correspond to positions to the right (left) for
    vertical boxes and above (below) for horizontal boxes

```python
from plotly import graph_objects as go
y = [1,14,14,15,19,20,20,20,25,28,28,29,35,50]

data = [go.Box(y=y,
              boxpoints='all',
              jitter=0.1,
              pointpos=0)]
fig = go.Figure(data=data)
fig.show()
```

### 5.2.4. Only display ouliers as points


```python
from plotly import graph_objects as go
y = [1,14,14,15,19,20,20,20,25,28,28,29,35,50]

data = [go.Box(y=y,
              boxpoints='outliers',
              jitter=0.1,
              pointpos=0)]
fig = go.Figure(data=data)
fig.show()
```

## 6.1. Basic Histogram<a class="anchor" id="27"></a>

```python
# Create a weighted list of numbers
lst = [x for x in range(0,50)]
lst1 = [x for x in range(25)]
lst2 = lst1[:][::-1]
weights = lst1 + lst2

import random
lst_hist = random.choices(lst, weights = weights, k = 500)

from plotly import graph_objects as go
data = [go.Histogram(x=lst_hist)]
layout = go.Layout(title='Basic Histogram')
fig = go.Figure(data=data,layout=layout)
fig.show()
```

## 6.2. Histogram - Change the start, end and the bin size<a class="anchor" id="28"></a>

```python
# Create a weighted list of numbers
lst = [x for x in range(0,50)]
lst1 = [x for x in range(25)]
lst2 = lst1[:][::-1]
weights = lst1 + lst2

import random
lst_hist = random.choices(lst, weights = weights, k = 500)

from plotly import graph_objects as go
data = [go.Histogram(x=lst_hist,name='List1',
                    xbins=dict(start=0,end=50,size=10))]
layout = go.Layout(title='Histogram - Change the start, end and the bin size')
fig = go.Figure(data=data,layout=layout)
fig.show()
```

# 8.2. Heatmap of weather data<a class="anchor" id="34"></a>

```python
import pandas as pd
df = pd.read_csv('../input/weather-dataset/weatherHistory.csv')
df = df.iloc[24:24*8] # Select 7 days

# Create a column with the day and one with the time
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'])
df['day'] = df['Formatted Date'].apply(lambda d: d.day)
df['hour'] = df['Formatted Date'].apply(lambda d: d.hour)

# Keep 3 columns
df = df[['day','hour','Temperature (C)']]
df.head()
```

```python
from plotly import graph_objects as go

data = [go.Heatmap(x=df['day'],
                   y=df['hour'],
                   z=df['Temperature (C)'].values.tolist(),
                   colorscale='Jet')
                   ]
layout = go.Layout(title='Temperatur by day and hour', 
                   xaxis={'title':'Day'},
                   yaxis={'title':'Time'})
fig = go.Figure(data=data, layout = layout)
fig.show()
```

# 9. Multiple Subplots<a class="anchor" id="35"></a>
## 9.1. Basic multiple Subplots<a class="anchor" id="36"></a>

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import numpy as np

# Generate random data
np.random.seed(0)
x1 = np.random.randint(1,101,100)
y1 = np.random.randint(1,101,100)

x2 = np.random.randint(1,150,100)
y2 = np.random.randint(1,150,100)

x3 = np.random.randint(1,50,100)
y3 = np.random.randint(1,50,100)

trace1 = go.Scatter(x=x1, y=y1, mode = 'markers')
trace2 = go.Scatter(x=x2, y=y2, mode = 'markers')
trace3 = go.Scatter(x=x3, y=y3, mode = 'markers')

fig = make_subplots(rows=1,cols=3)
fig.append_trace(trace1,1,1)
fig.append_trace(trace2,1,2)
fig.append_trace(trace3,1,3)
fig.show()
```

## 9.2. Multiple Subplots<a class="anchor" id="37"></a>

```python
import plotly.graph_objs as go
from plotly.subplots import make_subplots
import numpy as np

# Generate random data
np.random.seed(0)
x1 = np.random.randint(1,101,100)
y1 = np.random.randint(1,101,100)

x2 = np.random.randint(1,150,100)
y2 = np.random.randint(1,150,100)

x3 = np.random.randint(1,50,100)
y3 = np.random.randint(1,50,100)

trace1 = go.Scatter(x=x1, y=y1, mode = 'markers')
trace2 = go.Scatter(x=x2, y=y2, mode = 'markers')
trace3 = go.Scatter(x=x3, y=y3, mode = 'markers')

fig = make_subplots(rows=1,
                    cols=3,
                    shared_yaxes = True,
                    subplot_titles=['trace1','trace2','trace3'])

fig.append_trace(trace1,1,1)
fig.append_trace(trace2,1,2)
fig.append_trace(trace3,1,3)
fig.show()
```
