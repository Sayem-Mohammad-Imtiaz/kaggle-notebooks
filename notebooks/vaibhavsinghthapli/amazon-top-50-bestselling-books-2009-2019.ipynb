****Name of the Authors with highest number of published Books.****

```python
data["Author"].value_counts().sort_values(ascending=False).head(20).plot.bar()#color=color
plt.title("Top 20 Authors with highest number of Books Published (2009-2019)",fontsize=20)
plt.xlabel("Authors",fontsize=15)
plt.ylabel("No. of Books",fontsize=15)
plt.show()
```

As we see ("jeff kinney" published maximum books).

```python
plt.rcParams["figure.figsize" ]= (9,6)
#color = ["purple","yellowgreen"]#plt.cm.BuPu(np.linspace(0,1,15))
data["Genre"].value_counts().sort_values(ascending=False).head(15).plot.bar()#color=color
plt.title("No. of published Genre(2009-2019)",fontsize=20)
plt.xlabel("Genre",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

Under YEAR 2009 to 2019  Non-Fiction books are in trend.

**Lets see some data flow**

```python
data[["Name","User Rating"]].nlargest(20,"User Rating").plot.bar(x="Name",y="User Rating",width=0.4,)#'vertical', 'horizontal',color=color
plt.title("Top 20 Books with highest number of ratings",fontsize=20)
plt.xlabel("Name of the Books",fontsize=15)
plt.ylabel("No.of Ratings",fontsize=15)
plt.show()
```

```python
data[["Name","Reviews"]].nlargest(20,"Reviews").plot.bar(x="Name",y="Reviews",width=0.4,)#'vertical', 'horizontal',color=color
plt.title("Top 20 Books with highest number of Reviews",fontsize=20)
plt.xlabel("Name of the Books",fontsize=15)
plt.ylabel("No.of Reviews",fontsize=15)
plt.show()
```

As we see above I mentioned about reviews, now u also see that in above both graphs.

**Best selling Fiction books according to Reviews**

```python
data[data["Genre"] == "Fiction"].nlargest(10,"Reviews").plot.bar(x="Name",y="Reviews",width=0.4,)#'vertical', 'horizontal',color=color
plt.title("Top 10 bestselling Fiction books according to Reviews(2009-2019)",fontsize=20)
plt.xlabel("Name of the Books",fontsize=15)
plt.ylabel("No.of Reviews",fontsize=15)
plt.show()
```

**. So Top reviewed Fiction book is Where the Crawdads Sing with 87841 reviews**

**Best selling Non Fiction books according to Reviews**

```python
data[data["Genre"] == "Non Fiction"].nlargest(10,"Reviews").plot.bar(x="Name",y="Reviews",width=0.4,)#'vertical', 'horizontal',color=color
plt.title("Top 10 bestselling Non Fiction books according to Reviews(2009-2019)",fontsize=20)
plt.xlabel("Name of the Books",fontsize=15)
plt.ylabel("No.of Reviews",fontsize=15)
plt.show()
```

**So Top reviewed Non Fiction book is Becoming with 61133 reviews**


Most expensive Fiction books

```python
expensive = data[data["Genre"] == "Fiction"].nlargest(10,"Price")
p = expensive["Price"]
plt.rcParams["figure.figsize"] = (16,5)
#color = plt.cm.BuPu(np.linspace(0,1,20))
data[data["Genre"] == "Fiction"].nlargest(10,"Price").plot.bar(x="Name",y="Price",width=0.4,)#'vertical', 'horizontal',color=color
for i,price in enumerate(p):
    plt.text(i,price,str(price),fontsize=12)
plt.title("Top 10 Most expensive Fiction books(2009-2019)",fontsize=20)
plt.xlabel("Name of the Books",fontsize=15)
plt.ylabel("Price",fontsize=15)
plt.show()
```

**So The Twilight Saga is the most expensive Fiction Book**

Most expensive Non Fiction Books

```python
expensive = data[data["Genre"] == "Non Fiction"].nlargest(10,"Price")
p = expensive["Price"]
plt.rcParams["figure.figsize"] = (16,5)
#color = plt.cm.BuPu(np.linspace(0,1,20))
data[data["Genre"] == "Non Fiction"].nlargest(10,"Price").plot.bar(x="Name",y="Price",width=0.4,)#'vertical', 'horizontal',color=color
for i,price in enumerate(p):
    plt.text(i,price,str(price),fontsize=12)
plt.title("Top 10 Most expensive Non Fiction books(2009-2019)",fontsize=20)
plt.xlabel("Name of the Books",fontsize=15)
plt.ylabel("Price",fontsize=15)
plt.show()
```

**So Diagnostic and Statistical Manual of Mental Disorders,5th edition is the most expensive Non Fiction Book**

Top 15 Books according to price

```python
plt.rcParams["figure.figsize" ]= (16,5)
#color = plt.cm.PuRd(np.linspace(0,1,10))
data[["Name","Price"]].nlargest(10,"Price").plot.bar(x="Name",y="Price")#'vertical', 'horizontal',,color=color
plt.title("Top 10 Books with highest Price",fontsize=20)
plt.xlabel("Name of the Books",fontsize=15)
plt.ylabel("Price",fontsize=15)
plt.show()
```

```python
genre_reviews = data.groupby(["Genre","Year"],as_index=False)["Genre","Reviews"].mean()#,as_index=False
genre_reviews.head()
```

```python
plt.figure(figsize=(12,6))
plt.style.use("ggplot")
sns.barplot(x ="Year",y="Reviews",hue = "Genre",data = genre_reviews)
plt.title("No. of Reviews according to Year & Genre",Fontsize=20)
plt.show()
```

We can see that the sail of fiction books is always larger then Non-fiction books.
