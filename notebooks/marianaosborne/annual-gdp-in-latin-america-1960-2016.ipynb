```python
Happy = pd.read_csv('/kaggle/input/world-happiness-report-2021/world-happiness-report.csv')
Happy.head()

HaBR = Happy.loc[Happy["Country name"] == "Brazil"]
HaBR.plot(x="year", y=["Social support","Freedom to make life choices","Perceptions of corruption"])

HaBR["normHa"] = HaBR["Life Ladder"]/max(HaBR["Life Ladder"])
```

```python
data = pd.read_csv('/kaggle/input/petroleum-1983-2016/Datos histricos Futuros petrleo crudo WTI (1).csv',decimal=",")  # Para separação decimal com vírgula
data["Vol."]= data["Vol."].str.replace('M','') 
data["Vol."]= data["% var."].str.replace('%','') # Mudar string de colunas
data = data.rename(columns={"Vol.": "Vol. [M]", "% var.":"var. [%]"})        # Mudar nome de colunas
data[['Month','Year']] = data.Fecha.str.split(" ",expand=True)
dataO = data
dataO = dataO[['Month','Year', 'Último','Apertura', 'Máximo', 'Mínimo', "Vol. [M]", "var. [%]"]]

dataAp = dataO.loc[dataO.Month == "Abr"]   # Valores apenas de abril de cada ano
Petroleum = dataAp.astype({'Year': 'int32'})   # Mudar de objeto para int32 
Petroleum = Petroleum[['Year','Máximo', 'Mínimo']]
```
