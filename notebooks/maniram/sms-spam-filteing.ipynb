## Problem Description

In this given problem we have been given some text messages labele with spam or ham. We have to train
a model which can learn from it, and can predict message category to which it belongs to. 

```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
print(os.listdir("../input"))
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier,VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from wordcloud import WordCloud  # library to represent the words in text in pictorial format
```

## Data Preprocessing

There are some of the column in the data which are useless. so we have to firstly drop it

```python
data = pd.read_csv('../input/spam.csv',encoding='ISO-8859-1')
data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
data.head()
```

Convert the targetted column v1  into numerical format

```python
data['labels']=data['v1'].map({'ham':0,'spam':1})
data.head()
```

```python
X=data.v2
Y=data.labels
```

The data which are processed with ML model should be in numerical format. But in this data the messages are in 
textual format. So before processing it we have to convert in into numerical vector format. basically I will 
use two scikit-learn model i.e. CountVectorizer and TfIdfVectorizor to convert the text to vector format.
I will test each of them one by one and will choose that model which gives best accuracy.

```python
cv=CountVectorizer(decode_error='ignore')
X=cv.fit_transform(data['v2'])
trainX,testX,trainY,testY=train_test_split(X,Y,test_size=.3,random_state=42)
```

# ML Model Prediction

```python
model=MultinomialNB()
model.fit(trainX,trainY)
print(model.score(testX,testY))
```

The multinomial model gives us 97.8 % accuracy. Let us test svc model for it.

```python

dc=DecisionTreeClassifier()
dc.fit(trainX,trainY)
print(dc.score(testX,testY))
```

```python
rfc=RandomForestClassifier()
rfc.fit(trainX,trainY)
print(rfc.score(testX,testY))
```

```python
et=ExtraTreesClassifier()
et.fit(trainX,trainY)
print(et.score(testX,testY))
```

```python
knc=KNeighborsClassifier()
knc.fit(trainX,trainY)
print(knc.score(testX,testY))
```

In above models we are getting same level of accuracy as in CountVectorizor models. We can see that multinomial and other
decesion tree model are performing well so we will choose these model to predict the label of the message

```python
def visualize(label):
    words=''
    for msg in data[data['v1']==label]['v2'] :
        msg=msg.lower()
        words += msg + ' '

    wc=WordCloud(width=600,height=400).generate(words)
    plt.imshow(wc)
    plt.axis('off')
    plt.show()
```

Let us see what are the words which are playing major role in spam messages.

```python
visualize('spam')
```

Let us see what are the words which are playing major role in non-spam messages.

```python
visualize('ham')
```
