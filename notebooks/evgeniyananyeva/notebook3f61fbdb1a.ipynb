```python
df = df[["Hardness", "Solids", "Chloramines", "Conductivity", "Organic_carbon", "Turbidity", "Potability"]]
df.isnull().sum()
```

```python
df.head()
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(df.drop('Potability', axis=1), 
df['Potability'], test_size=0.2, random_state=42)
```

```python
train_x.shape
```

```python
test_x.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=13,class_weight = 'balanced_subsample')
model.fit(train_x, train_y)
```

```python
predicted = model.predict(test_x)
model.score(test_x, test_y)
```

```python
from sklearn.metrics import roc_auc_score
probabilities = model.predict_proba(test_x)
```

```python
roc_auc_score(test_y, probabilities[:, 1])
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(test_y, predicted)
```

```python
from sklearn import metrics
disp = metrics.plot_confusion_matrix(model, test_x, 
test_y,values_format = 'n',
display_labels= ["Potability"])
```

```python
from sklearn.metrics import precision_score

train_predictions = model.predict(train_x)
precision_score(train_y, train_predictions)
```

```python
from sklearn.metrics import recall_score

recall_score(train_y, train_predictions)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
from sklearn.metrics import roc_curve

fpr, tpr, _ = roc_curve(test_y, probabilities[:, 1])
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1], color='grey', lw=1, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
metrics.auc(fpr, tpr)
```
