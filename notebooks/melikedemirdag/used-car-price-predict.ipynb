```python
# We can see that 1970 has an outlier price mean. It may be a special car but it will negatively affect
# our model. That's why I prefer to delete
```

```python
newDataFrame = newDataFrame[newDataFrame.year != 1970]
newDataFrame
```

```python
newDataFrame.groupby("year").mean()["price"]
```

```python
newDataFrame.head()
```

```python
newDataFrame = newDataFrame.drop("transmission",axis =1)     # drop non numeric values

newDataFrame = newDataFrame.drop("model", axis =1) 
newDataFrame = newDataFrame.drop("fuelType", axis =1) 

newDataFrame
```

**MODEL**

```python
y  = newDataFrame["price"].values
x = newDataFrame.drop("price",axis = 1).values 
# because if we drop the price, only features will remain
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state = 10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python

scaler =MinMaxScaler ()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential ()
```

```python
model.add(Dense(12,activation ="relu"))
model.add(Dense(12,activation ="relu"))
model.add(Dense(12,activation ="relu"))
model.add(Dense(12,activation ="relu"))

model.add(Dense(1))

model.compile(optimizer ="adam", loss ="mse")
```

```python
model.fit(x =x_train, y = y_train , validation_data = (x_test,y_test) , batch_size = 250, epochs = 300  )
```

```python
lossData = pd.DataFrame(model.history.history)

lossData.head()
```

```python
lossData.plot()
```

```python
from sklearn.metrics import mean_squared_error , mean_absolute_error
```

```python
predictions = model.predict(x_test)
```

```python
predictions
```

```python
mean_absolute_error(y_test,predictions)      # mean difference between predict and actual value
```

```python
plt.scatter(y_test,predictions)
plt.plot (y_test,y_test,"r-*")
```
