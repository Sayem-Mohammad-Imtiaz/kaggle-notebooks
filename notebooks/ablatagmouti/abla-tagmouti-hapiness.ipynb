A negative value means that the variables are related in inverse relationship (if one increases the other decreses ). More the coefficient of correlation is close to -1 more the relation between the variables is strong.
As we can see the hapiness rank and the health (life expectency) are stongly and have an inverse relationship. The more the population have access to health the smallest the rank number is. As we can assume it is the same relation with the family and at a lesser scale with freedom. 

A positive coefficient means that the varaibles are linked in a positive relationship (if one increse the other too). the closer the coeff is from 1 the stronger the relationship is. if the coeff equal 0 the variables are not related.
As we can see the family the health access and the freedom have a strongly positive reliationship with the hapiness. the easier you have access to the health the more you have chance to be happy.

a higher number means a stronger relationship between the varaibles. As we can see the trust government has the lowest correlation with happiness. The impact on happiness is not as significant as for the others variable. .
