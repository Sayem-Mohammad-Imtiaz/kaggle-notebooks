# Logistic regression with significant variables (>0,4)

```python
# setting a series to output and a dataframe for the significant variables
y = df_heart['output']
X = df_heart[['cp', 'thalachh', 'exng', 'oldpeak']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# setting X and y for train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

lr_model = LogisticRegression().fit(X_train, y_train)
print(lr_model.score(X_test, y_test))
```

```python
scores = []
for i in [0.0001, 0.001, 0.01, 0.1, 1, 2, 5, 10]:
    lr_model = LogisticRegression(C = i).fit(X_train, y_train)
    scores.append(lr_model.score(X_test, y_test))

plt.figure(figsize = (12,5))
plt.plot(scores)
plt.title('Scores of Logistic Regression Classification - C = 0.0001 to 10', fontsize = '20')
plt.xlabel('C', fontsize = 15)
plt.ylabel('Score', fontsize = 15)
plt.xticks([0.0001, 0.001, 0.01, 0.1, 1, 2, 5, 10])

print('Max score: {:.4f}'.format(max(scores)))
```
