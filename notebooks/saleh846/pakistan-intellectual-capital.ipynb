```python
# importing important libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
original_data = pd.read_csv('/kaggle/input/pakistanintellectualcapitalcs/Pakistan Intellectual Capital - Computer Science - Ver 1.csv', encoding="latin1")
data_copy_1 = original_data.copy()  # making copy of data
data_copy_1.head()  # diplaying initial 5 rows
```

```python
# Basic information about data
data_copy_1.info()
```

```python
# Renaming columns names for convinience
data_copy_1 = data_copy_1.rename(columns={
    "Teacher Name": "teacher_name",
    "University Currently Teaching": "current_university",
    "Department": "department",
    "Province University Located": "province",
    "Designation": "designation",
    "Terminal Degree": "degree",
    "Graduated from": "graduated_from",
    "Country": "country",
    "Year": "year",
    "Area of Specialization/Research Interests": "specialization",
    "Other Information": "other_information"
})
data_copy_1.head()  # Displaying initial five rows
```

```python
# Droping "S#" column becauuse it is of no use as we have index
# Also dropping "other_information" column as it has almost all the null values
data_copy_1.drop(["S#", "other_information"], axis=1, inplace=True)
# Checking null values
data_copy_1.isnull().sum()
```

# **Task-1**
Which area of interest/expertise is in abundance in Pakistan and where we need more people?

### Where area of interest is in abundance? 

```python
specializations = pd.DataFrame(data_copy_1['specialization'])
specializations['specialization'].isnull().sum()  # Number of null values in 'specialization' column
```

```python
specializations.dropna(inplace=True) # Dropping the rows with null values
```

```python
specializations['specialization'] = specializations.specialization.str.lower()  # converting into lower case
specializations['specialization'] = specializations.specialization.str.replace('.', '') # removing all periods '.'
specializations['specialization'] = specializations.specialization.str.replace('and ', '') # removing all the 'and '
# As majority of the records in this cloumn have multiple areas of interets separated by commas ',' so splitting the records on the basis of commas ','
specializations['specialization'] = specializations.specialization.str.split(',')
```

```python
# making a list which contains areas of interests individually
lst = []
for i in specializations['specialization']:
    for j in i:
        lst.append(j.strip())

# making a dataframe which contains the areas of interests individually in rows
df = pd.DataFrame(lst)
```

```python
df = df.rename(columns={
    0: 'area_of_interest'
})
task1_frame = pd.DataFrame(df.area_of_interest.value_counts())
```

```python
# reseting tthe index
task1_frame = task1_frame.reset_index()
# changing columns names
task1_frame = task1_frame.rename(columns={
    'index': 'area_of_interest',
    'area_of_interest': 'count'
})
```

```python
# plotting top 15 areas of interests
plt.figure(figsize=(15, 8))
sns.barplot(x=task1_frame.loc[0:14, 'area_of_interest'], y=task1_frame.loc[0:14, 'count'])
plt.xticks(rotation=90, fontsize=15)
plt.yticks(fontsize=15)
plt.xlabel("Areas of Interest", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.title("Top Areas of Interests", fontsize=20)
plt.show()
```

**From this plot we can see "software engineering" and "machine learning" are the top areas of interest which are specifically mentioned**

### Where we need more professors? 

```python
data_science = 0
for i in task1_frame['area_of_interest']:
    if 'data science' in i:
        data_science += 1
print("Data Science:", data_science)
cyber_security = 0
for i in task1_frame['area_of_interest']:
    if 'cyber security' in i:
        cyber_security += 1
print("Cyber Security:", cyber_security)
```

**By analzing the data we get to see that there are many important fields in which we have very less number of professors such as "data science" which is mentioned by only 4 professors as thier area of interest and "cyber security" which is mentioned by only 1 professors as thier area of interest**
