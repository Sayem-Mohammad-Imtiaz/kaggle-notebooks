```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# Categorical Attributes

```python

fig, ax = plt.subplots(2,2,figsize=(15, 10))
sns.countplot(y=df['gender'],ax=ax[0,0])
sns.countplot(y=df['hypertension'],ax=ax[0,1])
sns.countplot(y=df['heart_disease'],ax=ax[1,0])
sns.countplot(y=df['ever_married'],ax=ax[1,1])
```

# Nominal Attributes

```python
fig, ax = plt.subplots(1,3,figsize=(30, 10))
sns.countplot(y=df['smoking_status'],ax=ax[0])
sns.countplot(y=df['Residence_type'],ax=ax[1])
sns.countplot(y=df['work_type'],ax=ax[2])
```

# Numerical Values

```python
fig, ax = plt.subplots(1,3,figsize=(30, 10))
sns.distplot(df['age'],ax=ax[0])
sns.distplot(df['avg_glucose_level'],ax=ax[1])
sns.distplot(df['bmi'],ax=ax[2])
```

First Normal distribution , second normal with right tail,third normal with some outliers.
