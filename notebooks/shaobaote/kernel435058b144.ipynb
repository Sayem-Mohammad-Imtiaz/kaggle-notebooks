```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as dates
import networkx as nx
import re
import time
import numpy as np

# get data and remove the NA
csv_data_users = pd.read_csv('../input/meta-kaggle/Users.csv')  
print(csv_data_users.shape)
csv_data_users.dropna(axis=0, how='any', inplace=True)
print(csv_data_users.shape)
print(csv_data_users.info())
```

```python
#generate the dict including the year & month
date={}
year=2010
while True:
    months=[]
    for x in range(12):
        months.append(x+1)
        year=year+(1 if((x+1)%12==0) else 0)
    date.update({year-1: months})
    if year==2019:
        break    
print(date)

#generate a dict to stock the num of registered men 
date_count={}
year=2010
while True:
    months=[]
    for x in range(12):
        month_count={x+1:0}
        months.append(month_count)
        year=year+(1 if((x+1)%12==0) else 0)
    date_count.update({year-1: months})
    if year==2019:
        break    
print(date_count)

pattern=re.compile('(.*)/(.*)/(.*)')

#count the total registered men in per month from 2010(1/1)-2018(12/10)
def func(b,limit):
    a=b
    for i in range((limit-1)*500000+1,limit*500000+1):
        if i>2412167: return a
        m=pattern.match(csv_data_users.iloc[i,3])
        print(m)
        for x,y in date.items():
            for z in y:
                if (m.group(3)==str(x)and(m.group(1)==str(z))):
                    a[x][y.index(z)][z]=a[x][y.index(z)][z]+1
                    print(a[x][y.index(z)])
    return a

#such a big dataset can't be done stright,so dividing it to 5 parts and sum up all of the counts in the end
#you can just run a part one time
date1=func(date_count,1)
print(date1)
date2=func(date_count,2)
print(date2)
date3=func(date_count,3)
print(date3)
date4=func(date_count,4)
print(date4)
date5=func(date_count,5)
print(date5)
```
