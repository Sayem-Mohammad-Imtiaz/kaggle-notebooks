```python
# col_names = ["Permit Type", "Permit Creation Date", "Block", "Lot", "Street Numbe", "Existing Construction Type", "Supervisor District", "Zipcode"]

raw_df = pd.read_csv("/kaggle/input/building-permit-applications-data/Building_Permits.csv", header=None)
```

```python
col_names = {i:raw_df.iloc[0][i] for i in range(len(raw_df.iloc[0]))}
col_indices = {raw_df.iloc[0][i]:i for i in range(len(raw_df.iloc[0]))}
df = raw_df.rename(col_names,axis='columns').iloc[1:]
```

```python
feature_cols = ["Estimated Cost", "Revised Cost"]
all_cols = feature_cols + ["Filed Date", "Issued Date"]
df = df[all_cols]
df = df.dropna()
```

```python
df['Filed Date'] = pd.to_datetime(df['Filed Date'])
df['Issued Date'] = pd.to_datetime(df['Issued Date'])
```

```python
df['time'] = ((df['Issued Date'] - df['Filed Date'])/np.timedelta64(1, 'D')).astype(int)
```

```python
# Split dataset into training set and test set

X = df[feature_cols]
y = df['time']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3) # 70% training and 30% test
```

```python
X,y
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
