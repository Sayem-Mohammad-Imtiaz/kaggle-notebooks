# Create a global map for COVID Vaccinations
The data I chose to present is 'people_vaccinated_per_hundred' which is the number of vaccinated people per hundred, regardless of the completion of the vaccinations. Although this number does not present the real number of people who start to get the effect of the vaccinations, I personally think that it could show the overall picture of the countries' situations better. If we count only people who are fully vaccinated, it means we completely ignore the country who just start giving the first dose of vaccines to people. With this method, we can differentiate the countries that just starting handling out vaccines from countries that have no vaccines at all. On the other hand, we cannot differ between the number of the first dose and the second dose. In my opinion, this is fine because I think our top priority should be to help countries which do not have vaccinations at all or have significantly lower number of overall vaccinations.

```python
import numpy as np
import pandas as pd
import plotly.express as px
#import plotly
df=pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv')
df.index= pd.to_datetime(df['date'])
df=df.sort_index()
print(df.columns)
#print(df)
fig=px.choropleth(df,locations = 'country',locationmode='country names',color='people_vaccinated_per_hundred',hover_name='country',animation_frame='date',color_continuous_scale='rainbow',range_color=(0.0,60.0))
fig.update_layout(title_text="Percentage of the country's total population that are vaccinated at least once" ,title_x=0.42)
fig.show()

#plotly.offline.plot(fig,filename='vaccinations.html')
```

# Create a global map of earthquake occurrences worldwide
The severe earthquake I choose is the earthquake that happened near Kuril Islands, Russia on March 25. The magnitude of it is 7.5 which ranks in fourth place in the ranking of global earthquakes with highest magnitudes in 2020. The earthquake with this magnitude can cause tsunamis far away from its epicentre. Therefore, at first the warning are sent to Russia, Japan, and adjacent islands. However, nothing happened and the warning was lifted afterward. No casualty was reported.

From the animation, the earthquake happened at 2.00 am of March 25. The earthquake became weaker and persisted until 4.00 am. We can observe that there are few other weak earthquakes happen in the same area throughout the day. This may be the impact from the first major earthquake.
