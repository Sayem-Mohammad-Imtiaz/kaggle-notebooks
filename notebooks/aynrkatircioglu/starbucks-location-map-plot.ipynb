```python
starbucks = pd.read_csv("/kaggle/input/store-locations/directory.csv")
df = starbucks.copy()
df.head()
```

```python
df["color"]="rgb(0,112,74)" 
```

```python
data = [dict(
    type='scattergeo',
    lon = df['Longitude'],
    lat = df['Latitude'],
    hoverinfo = 'text',
    text = "Country: "+df["Country"] +"City: " + df.City + " Street Adress: "+df["Street Address"]+" Store Name: " + df['Store Name'],
    mode = 'markers',
    marker=dict(
        sizemode = 'area',
        sizeref = 1,
        size= 10 ,
        line = dict(width=1,color = "rosybrown"),
        color = df["color"],
        symbol = 'circle-dot',
        opacity = 0.8),
)]
layout = dict(
    title = 'Starbucks Locations Around The World',
    hovermode='closest',
    geo = dict(showframe=True, showland=True, showcoastlines=True, showcountries=True,
               countrywidth=1, projection=dict(type='miller'),
              landcolor = 'rgb(240, 213, 185)',
              subunitwidth=1,
              showlakes = True,
              lakecolor = 'blue',
              countrycolor="rgb(5, 5, 5)")
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
