```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
import random
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
num_bins = 10
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report, plot_confusion_matrix
```

IMPORTING DATASET

```python
pima_df = pd.read_csv('/kaggle/input/pima-indians-diabetes-database/diabetes.csv')
pima_df
```

RENAMING COULUMNS FOR CONVENIENCE

```python
colnames = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
pima_df = pd.read_csv('/kaggle/input/pima-indians-diabetes-database/diabetes.csv', names = colnames, header = 0)
pima_df
```

DESCRIBING THE DATASET

```python
pima_df.info()
print(pima_df.shape)
```

```python
pima_df.describe()
```

Replacing zeros with median values of the respective series wherever applicable..

```python
pima_df.loc[pima_df.plas == 0, 'plas'] = pima_df.plas.median()
pima_df.loc[pima_df.pres == 0, 'pres'] = pima_df.pres.median()
pima_df.loc[pima_df.skin == 0, 'skin'] = pima_df.skin.median()
pima_df.loc[pima_df.test == 0, 'test'] = pima_df.test.median()
pima_df.loc[pima_df.mass == 0, 'mass'] = pima_df.mass.median()
```

```python
pima_df.describe()
```

EVALUATING THE CLASSES

```python
pima_df.groupby("class").agg({'class': 'count'})
```

```python
sns.countplot(data = pima_df,x = "class")
```

PAIRPLOT OF ATTRIBUTES

```python
sns.pairplot(data = pima_df, hue = 'class',palette = 'bright')
```

SPLITTING THE DATA IN TO TRAIN AND TEST

```python
n=pima_df['class'].count()
train_set = pima_df.head(int(round(n*0.7)))
test_set = pima_df.tail(int(round(n*0.3)))
test_set
```

```python
train_labels = train_set.pop("class")
test_labels = test_set.pop("class")
```

APPLYING DECISION TREE CLASSIFIER

```python
#DecisionTreeClassifier
dtmod = DecisionTreeClassifier(criterion='entropy')
dtmod.fit(train_set, train_labels)
```

```python
test_pred = dtmod.predict(test_set)
metrics.accuracy_score(test_pred, test_labels)
```

```python
cm=metrics.confusion_matrix(test_labels, test_pred, labels=[1, 0])
dtdia = pd.DataFrame(cm, index = [i for i in ["1","0"]],columns = [i for i in ["Predict 1","Predict 0"]])
plt.figure(figsize = (7,5))
sns.heatmap(dtdia, annot=True)
print("CONFUSION MATRIX FOR LINEAR MODEL:")
print(classification_report(test_labels, test_pred))
```
