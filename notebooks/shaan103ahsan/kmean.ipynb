```python
df = pd.read_csv("../input/clicks-conversion-tracking/KAG_conversion_data.csv")
df.head()
```

```python
print(df.xyz_campaign_id.unique())
```

```python
y_def=df.iloc[:,[1]].values
```

```python
x=df.iloc[:, [6, 8]].values
x[0:5]
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss, c = 'blue')
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
print(y_kmeans)
```

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = '916')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'green', label = '936')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'black', label = '1178')

#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'blue', label = 'Centroids')

plt.legend()
```

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = '916')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'green', label = '936')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 100, c = 'black', label = '1178')
#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 100, c = 'blue', label = 'Centroids')
plt.show()
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
score = silhouette_score(x, y_kmeans, metric='euclidean')
print('Silhouetter Score: %.3f' % score)
```

```python

print(kmeans.labels_)
y=list(np.concatenate(y_def).flat)
print(y_def)
```

```python
from sklearn.metrics import adjusted_rand_score
ari_kmeans = adjusted_rand_score(y, kmeans.labels_)
print("ARI Score : ",ari_kmeans)
```

```python
from sklearn.metrics.cluster import normalized_mutual_info_score
nmi_kmeans = normalized_mutual_info_score(y, kmeans.labels_)
print("NMI Score : ",nmi_kmeans)
```
