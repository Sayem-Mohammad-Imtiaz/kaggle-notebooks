```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

Objective : To predict selling price of a car based on given information

```python
data=pd.read_csv("/kaggle/input/vehicle-dataset-from-cardekho/CAR DETAILS FROM CAR DEKHO.csv")
data.head()
```

Adding a new variable called 'Age', this will indicate how old the car is

```python
data['age']=2021-data['year']
data.head()
```

We do not required model name or year so we can drop these two variables

```python
data=data.drop(['name','year'],axis=1)
data.head()
```

```python
data.shape
```

**EDA** - Exploratory Data Analysis help to understand data better. We can analyse data with the help of charts, graphs and check the relationship between different variables used for model. 

We will start with Describe() that gives basis statistics for all numeric columns

```python
data.describe()
```

```python
#let's check if there are null values, duplicate values in the give dataset. 
#Also, we will convert or use dummies for the categorical variables moving forward
data.isnull().sum()
```

```python
data.duplicated().sum()
# We have 842 duplicate records and we can remove them to clean our dataset. 
#This is really important step to check as having duplicate values can impact your prediction
```

```python
data.drop_duplicates(subset=None, inplace=True)
data.shape
```

```python
#Creating Dummy Variables or converting categorical variables into numeric values for better analysis
#I am using LabelEncoder from the sklearn.preprocessing
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data.head()
data['fuel']=le.fit_transform(data['fuel'])
data['seller_type']=le.fit_transform(data['seller_type'])
data['owner']=le.fit_transform(data['owner'])
data['transmission']=le.fit_transform(data['transmission'])
```

```python
data.head()
```

```python
#we can use unique() function to check values of each converted columns
print(data['fuel'].unique())
print(data['owner'].unique())
print(data['seller_type'].unique())
print(data['transmission'].unique())
```

Building a Linear Model
> We will split dataset into two parts - train and test. First we will use train dataset and let the model find the best linear line. Then we will check with test model that how much prediction is accurate

```python
x=pd.DataFrame(data,columns=['km_driven','fuel','seller_type','transmission','owner','age'])
y=pd.DataFrame(data,columns=['selling_price'])
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
print("intercept :",reg.intercept_)
print("slope :",reg.coef_)
```

```python
y_pred=reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
score=r2_score(y_test,y_pred)
print(score)
```

```python
#it has an accuracy of 36%
```
