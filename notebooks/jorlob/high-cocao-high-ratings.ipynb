# High Cocao % = High Ratings?

Quick visual EDA of the dataset to see assess if there is a any evidence of the higher the percentage of cacoa the higher the review rating.

Import the needed packages

```python
import pandas as pd

import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
df = pd.read_csv('../input/flavors_of_cacao.csv',  na_values = [" ",'Ã‚ '])
```

Check of the columns and names of columns

```python
df.columns
```

As you can see the column names are not clear, so lets reasign new names

```python
df.columns = ['Company', 'Specific_Bean_Origin', 'REF', 'Review_Date', 'Cocoa_Percent', 'Company_Location', 'Rating', 'Bean_Type', 'Broad_Bean_Org']
```

Get a overview of the data types and then how the data structured.

```python
df.info()
```

```python
df.head()
```

In order to manipulate some of the data we need to change some column datatypes

Company to catergoy and Review_Date to numeric.

Cocoa_Percent needs to be striped of its '%' and then converted to numeric.

```python
df['Company'] = df['Company'].astype('category')
df['Review_Date'] = pd.to_numeric(df['Review_Date'])
df['Cocoa_Percent']= df['Cocoa_Percent'].str.strip('%').astype(float)
```

## Univariate Analysis

Create some histograms to access the distrubtions of Ratings and Cocoa_Percentage

```python
df.describe()
```

We can assess that the max rating was a 5.0 while the avg cocao percent was 71.698 with a max of 100.

```python
df['Rating'].plot(kind = 'hist')
plt.title('Histogram of Ratings')
plt.xlabel('Rating')
plt.ylabel('Count of Ratings')
plt.show()

df['Cocoa_Percent'].plot(kind = 'hist')
plt.title('Histogram of Cocoa_Percent')
plt.xlabel('Cocoa_Percent')
plt.ylabel('Count')
plt.show()
```

We can tell that the majortiy of ratings fall between 3.0 and 3.5, while the majority of chocolate bars had a cocao percentage of 65to 75.

## Bivariate Analysis

Explore if there is any relationship between the rating and another metric

```python
df.plot(kind = 'scatter', x = 'Review_Date', y = 'Rating', rot=70)
plt.xlabel('Year of Review')
plt.title('Scatterplot of Review Year and Rating')
plt.show()
```

The data shows that the ratings for the chocolatebars has narowed over the years. This could inidcate that the reviewers had become more discerning about quality or that the quality of the chocolate bars had been standardised by 2014.

The highest and lowest  ratings were given from 2006 to 2008. They are two 5.0 ratings and three 1.0 ratings.

```python
df.plot(kind = 'scatter', x = 'Rating', y = 'Cocoa_Percent', rot=70)
plt.ylabel('Cacoa %')
plt.title('Scatterplot of Rating and Cacoa %')
plt.show()
```

The above scatterplot shows that having a cocao percentage of 100 does not guarantee a high rating. The 100% chocolatebars had a max rating of 3.5 and min rating 1.0.

This gets confirmed when calculating the correlation between the two columns.
