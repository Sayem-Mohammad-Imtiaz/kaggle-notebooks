```python
msk = np.random.rand(len(df)) < 0.8
train = df_p[msk]
test = df_p[~msk]
```

```python
from sklearn import linear_model
```

```python
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['Close','Low','High']])
train_y = np.asanyarray(train[['VWAP']])
regr.fit (train_x, train_y)
```

```python
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(train.Close, train.Low, train.High,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.coef_[0][1]*train_x + regr.coef_[0][2]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Close,Low,High")
plt.ylabel("VWAP")
```

```python
from sklearn.metrics import r2_score
```

```python
test_x = np.asanyarray(test[['Close','Low','High']])
test_y = np.asanyarray(test[['VWAP']])
test_y_ = regr.predict(test_x)
```

```python
print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y , test_y_) )
```

```python
print("Residual sum of squares: %.2f"
      % np.mean((test_y_ - test_y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(test_x, test_y))
```
