```python
data=pd.read_csv('/kaggle/input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
print(data.shape,data.columns,sep='----')
```

```python
columns=data.columns.tolist()
for x in columns:
    if isinstance(data.loc[0,x],str):
        print('columns '+str(x)+' categorical label')
```

```python
print(data.head())
```

```python
import matplotlib.pyplot as plt
data['score']=data['User Rating']*data['Reviews']
data['score'].plot(kind='box')
plt.yticks(rotation=45)
```

```python
plt.style.use('ggplot')
data[data['score']>10000]['Genre'].value_counts().plot.bar()
plt.xticks(rotation=0)
```

```python
zuozhe=data[data['score']>10000]['Author'].value_counts()
zuozhe[zuozhe>8].plot.bar()
plt.xticks(rotation=45)
```
