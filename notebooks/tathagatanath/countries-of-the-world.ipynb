```python
# removing additional space in country and region names 

for i in range(data_copy.shape[0]) :
    
    data_copy.at[i,'Country'] = data_copy.at[i,'Country'].strip()
    data_copy.at[i,'Region'] = data_copy.at[i,'Region'].strip()
```

```python
# renaming countries

c={7:'Antigua and Barbuda', 14:'The Bahamas', 25:'Bosnia and Herzegovina', 28:'British Virgin Islands', 39:'Central African Republic',
                  45:'Democratic Republic of the Congo', 46:'Republic of the Congo', 73:'The Gambia', 109:'North Korea',
                  110:'South Korea', 136: 'Federated States of Micronesia', 153:'Northern Mariana Islands',
                  178:'São Tomé and Príncipe', 174:'St Pierre and Miquelon', 204:'Trinidad and Tobago',
                  208:'Turks and Caicos Islands'}

for i in c :
    data_copy.at[i,'Country']=c[i]
```

```python
data_copy.iloc[list(c.keys()),:]
```

```python
# replacing coma(,) with dot(.)

x=data_copy.columns.tolist()

for i in x[4:8]+x[9:] :
    
    for j in range(data_copy.shape[0]) :
        
        if str(data_copy.at[j,i]) != 'nan' :
            
            if ',' in str(data_copy.at[j,i]) :
                
                temp = str(data_copy.at[j,i]).split(',')
                
                data_copy.at[j,i] = temp[0]+'.'+temp[1]
```

```python
data_copy=data_copy.astype({'Pop. Density (per sq. mi.)':float, 'Coastline (coast/area ratio)':float, 'Net migration':float, 'Infant mortality (per 1000 births)':float, 
                            'GDP ($ per capita)':float, 'Literacy (%)':float, 'Phones (per 1000)':float, 'Arable (%)':float, 'Crops (%)':float, 'Other (%)':float,
                               'Climate':float, 'Birthrate':float, 'Deathrate':float, 'Agriculture':float, 'Industry':float, 'Service':float})
```

```python
data_copy.info()
```

```python
data_copy.isnull().sum()
```

```python
# filling nan values with the mean of that column for the region in which the country is

nan_columns=data_copy.columns.tolist()[6:]

for i in range(data_copy.shape[0]) :
    
    for j in nan_columns :
        
        if str(data_copy.at[i,j]) == 'nan'  :
            
            region=data_copy.at[i,'Region']
            
            temp_data=data_copy[data_copy['Region']==region]
            
            temp_mean=round(temp_data[j].mean(),2)
            
            data_copy.at[i,j]=temp_mean
```
