```python
# importing required libraries

# import numpy as np
# import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_absolute_error
```

# Loading and Analysing the Dataset

```python
# loading the dataset

data = pd.read_csv('/kaggle/input/housesalesprediction/kc_house_data.csv')
data.head()
```

```python
data.columns
```

```python
dCol = ['id', 'date', 'zipcode']
data.drop(dCol, axis=1, inplace=True)
data.head()
```

```python
data.shape
```

```python
data.describe()
```

# Visualization of Features

```python
# plotting the relation of various fields

fig, ax = plt.subplots(5,3, sharey=True, figsize=(25,20))

ind = 1
for i in range(5):
    ax[i][0].set_ylabel('price')
    for j in range(3):
        if ind == 14:
            ind += 2
        ax[i][j].scatter(data.iloc[:, ind], data['price'])
        ax[i][j].set_xlabel(data.columns[ind])
        ind += 1
```

# Data Preprocessing

```python
# checking presence of missing data

data.isnull().sum()
```

```python
# abstraction of features and label

X, y = data.drop('price', axis=1), data['price']
print(X.shape, y.shape)
```

```python
# cross validation scores

cross_val_score(RandomForestRegressor(n_estimators=100), X, y, cv=5)
```

```python
# splitting the dataset into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
print(y_train.shape, y_test.shape)
```

```python
# scaling the features in order to avoid bias of features

st = StandardScaler()
X_train = st.fit_transform(X_train)
X_test = st.transform(X_test)

X_train[:2]
```

# Applying Regression Algorithm

```python
# fitting the data into the model

reg = RandomForestRegressor(n_estimators=100).fit(X_train, y_train)
print('Training Data Score: {:.2f}'.format(reg.score(X_train, y_train)))
print('Testing Data Score: {:.2f}'.format(reg.score(X_test, y_test)))
```

```python
reg.feature_importances_
```

# Model Evaluation

```python
# R2-Score of the model and mean_absolute_error

print('R2 Score: {}'.format(r2_score(y_test, reg.predict(X_test))))
print('Mean Absolute Error: {}'.format(mean_absolute_error(reg.predict(X_test),y_test)))
```
