# Seggregating the grades based on marks scored

```python
# Create three empty series 

math_grade = []
reading_grade=[]
writing_grade=[]

# Create a function to seggregate the grades based on marks

def grading(score,grade):
    for i in score:
        if (i>90):
            grade.append('A')
        elif (i>80):
            grade.append('B')
        elif (i>65):
            grade.append('C')
        elif (i>50):
            grade.append('D')
        elif (i>40):
            grade.append('E')
        else:
            grade.append('Fail')
            
"""Enter the fucntion with arguments of specific column and a new series name"""
```

```python
# Use the function to create three series of Grades for corresponding marks for the subjects

grading(df['math score'],math_grade)
grading(df['reading score'],reading_grade)
grading(df['writing score'],writing_grade)

# Create three new columns in the dataframe using these series  

df['math_grade'] = math_grade
df['reading_grade'] = reading_grade
df['writing_grade'] = writing_grade
```

**> Grades achieved in each skill**

```python
# Create a new series with the count of students falling under each grade for corresponding subjects

m_g = df['math_grade'].groupby(df['math_grade']).count()
r_g = df['reading_grade'].groupby(df['reading_grade']).count()
w_g = df['writing_grade'].groupby(df['writing_grade']).count()
```

```python
fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize = (12,12))
ax1.set_title('MATH - GRADE')
ax2.set_title('READING - GRADE')
ax3.set_title('WRITING - GRADE')
 
labels = m_g.keys()
ax1.pie(m_g, autopct="%.1f%%",labels = labels)
ax2.pie(r_g, autopct="%.1f%%",labels = labels)
ax3.pie(w_g, autopct="%.1f%%",labels = labels)

fig.savefig('grades.jpeg')
```

# Machine learning - Linear Regression

```python
x = df[['math score','reading score','writing score']]
y = df['test preparation course']
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 45)

print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

# Define model. Specify a number for random_state to ensure same results each run
melbourne_model = LogisticRegression(random_state=1)

# Fit model
melbourne_model.fit(x_train, y_train)
```

```python
y_pred = melbourne_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

# creating a confusion matrix
cm = confusion_matrix(y_test, y_pred)
```
