```python
#check the unique values 

print(df.cut.unique())
```

* Above cut type is reference of diamond cutting. There are 5 types of cutting. 

```python
df.cut.value_counts().plot(kind='bar', color='green')

plt.text(x = -0.15, y = df.cut.value_counts()[1]+1, s = df.cut.value_counts()[1])
plt.text(x = 0.85, y = df.cut.value_counts()[0]+2, s = df.cut.value_counts() [0])
plt.text(x = 1.85, y = df.cut.value_counts()[2]+3, s = df.cut.value_counts() [2])
plt.text(x = 2.85, y = df.cut.value_counts()[3]+4, s = df.cut.value_counts() [3])
plt.text(x = 3.85, y = df.cut.value_counts()[4]+5, s = df.cut.value_counts() [4])

plt.title('Count the value cutting edge', fontsize=18)
plt.xlabel('Cut', fontsize=15, fontweight='bold', fontstyle='italic')
plt.ylabel('Counts', fontsize=15, fontweight='bold', fontstyle='italic')

plt.show()
```

**Box Plot**

```python
print('The minimum value of diamond price:',(df['price'].min()))

print('The maximum value of diamond price:', (df['price'].max()))
```

```python
plt.boxplot(df['price'])

#add title and label
plt.title('Box Plot of Price', fontsize=15)
plt.ylabel('Count', fontsize=12)

#add labesl of five summary 
plt.text(x = 1.1, y=df['price'].min(), s='min')
plt.text(x = 1.1, y=df['price'].max(), s='max')

plt.text(x = 1.1, y=df.price.quantile(0.25), s='Q1')
plt.text(x = 1.1, y=df['price'].median(), s='median(Q2)')
plt.text(x = 1.1, y=df.price.quantile(0.75), s='Q3')


#check outliers
#add an arrow(annotate) to show the outliers



plt.show()
```
