This data set consists of the marks secured by the students in various subjects.

* **gender:** female,male

* **race/ethnicity:** group C,group D

* **parental level of education:** some college, associate's degree

* **lunch:** standard, free/reduced

* **test preparation course:** none, completed

* **math score:** subject marks

* **reading score:** How much time spend on reading

* **writing score:** Writing speed 

```python
#import required library

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
dataset = pd.read_csv("../input/students-performance-in-exams/StudentsPerformance.csv")
```

```python
#check top 7 exampes

dataset.head(7)
```

```python
#IS there any null values?

dataset.isnull().sum()
```

```python
# Ploting the distribution of gender
labels=dataset['gender'].value_counts().index
values=dataset['gender'].value_counts().values

plt.figure(figsize=(6,6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Gender')
plt.show()
```

```python
# Ploting the distribution of lunch
labels=dataset['lunch'].value_counts().index
values=dataset['lunch'].value_counts().values

plt.figure(figsize=(6,6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Lunch')
plt.show()
```

> From above graph, Standard lunch is common with 64.5% more than free/reduced lunch with 35.5%.

**Box Plot**

```python
plt.boxplot(dataset['math score'])

#add title
plt.xlabel('Box Plot of math score', fontsize=16)
plt.ylabel('Math Score', fontsize=16)


#summary 
plt.text(x = 1.1, y=dataset['math score'].min(), s='min')
plt.text(x = 1.1, y=dataset['math score'].max(), s='max')

plt.text(x = 1.1, y=dataset['math score'].median(), s='median(Q2)')
#plt.text(x = 1.1, y=dataset.math score.quantile(0.25), s='Q1')
#plt.text(x = 1.1, y=dataset.math score.quantile(0.75), s='Q3')

plt.annotate('Outliers', xy=(0.97, 22), xytext=(0.7, 44), arrowprops=dict(facecolor='blue', arrowstyle='simple'))

plt.show()
```
