```python
import numpy as np
import pandas as pd
import plotly.express as px
import matplotlib.pyplot as plt

%matplotlib inline

df = pd.read_csv("../input/ibm-hr-analytics-attrition-dataset/WA_Fn-UseC_-HR-Employee-Attrition.csv")
```

```python
df.Attrition.value_counts().plot.bar()
```

```python
df.info()
```

```python
edu_attrition_income = df[["Education", "Attrition", "MonthlyIncome"]]
edu_attrition_income = edu_attrition_income.replace({"Education": {1:'Below College',2:'College',3:'Bachelor',4:'Master',5:'Doctor'}})
edu_attrition_income = edu_attrition_income.groupby(["Education", "Attrition"]).agg('mean')
print(edu_attrition_income)
edu_attrition_income.unstack(level=-1)['MonthlyIncome'].plot(kind='bar')
```

```python
field_attrition_income = df[["EducationField", "Attrition", "MonthlyIncome"]]
field_attrition_income = field_attrition_income.groupby(["EducationField", "Attrition"]).agg('mean')
print(field_attrition_income)
field_attrition_income.unstack(level=-1)['MonthlyIncome'].plot(kind='bar')
```

```python
df[["PercentSalaryHike", "Attrition"]].groupby(["Attrition"]).agg('mean').plot.bar()
```

```python
year_attrition = df[["YearsAtCompany", "Attrition"]].groupby(["YearsAtCompany","Attrition"]).size().unstack(level=-1).fillna(0)
year_attrition.plot.line()
```

```python
year_attrition["percent"] = year_attrition["Yes"] * 100 / (year_attrition["Yes"] + year_attrition["No"])
year_attrition["percent"].plot.line()
```

```python
values = {1:"Low", 2:"Medium", 3:"High", 4:"Very High"}
invol_satis = df[["JobInvolvement","JobSatisfaction", "Attrition"]].replace({"JobInvolvement": values, "JobSatisfaction": values})
```

```python
invol_satis[invol_satis.Attrition == "Yes"].groupby(["JobInvolvement","JobSatisfaction"]).size().unstack(level=-2) * 100 / df.Attrition.value_counts().Yes
```

```python
invol_satis[invol_satis.Attrition == "No"].groupby(["JobInvolvement","JobSatisfaction"]).size().unstack(level=-2) * 100 / df.Attrition.value_counts().No
```

```python
gender_attrition_income = df[["Gender", "Attrition", "MonthlyIncome"]]
gender_attrition_income = gender_attrition_income.groupby(["Gender", "Attrition"]).agg('mean')
print(gender_attrition_income)
gender_attrition_income.unstack(level=-1)['MonthlyIncome'].plot(kind='bar')
```

```python
gender_attrition_income = df[["Gender", "Attrition", "MonthlyIncome"]]
fig = px.box(gender_attrition_income, x="Gender", y="MonthlyIncome", points="all", color="Attrition")
fig.show()
```

```python
comp_attrition = df[["NumCompaniesWorked", "Attrition"]].groupby(["NumCompaniesWorked","Attrition"]).size().unstack(level=-1).fillna(0)
comp_attrition.plot.line()
```

```python
comp_attrition["percent"] = comp_attrition["Yes"] * 100 / (comp_attrition["Yes"] + comp_attrition["No"])
comp_attrition["percent"].plot.line()
```
