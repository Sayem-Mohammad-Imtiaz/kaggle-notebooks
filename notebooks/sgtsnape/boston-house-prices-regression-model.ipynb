```python
#DATA

file_path='/kaggle/input/boston-house-prices/housing.csv'
heading=['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','PRICE']
house_data=pd.read_csv(file_path,delimiter='\s+',names=heading)

#print(house_data.shape)
X=house_data[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']]
y=house_data.PRICE

```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split as TTS

#LINEAR REGRESSION

model=LinearRegression()

X_train,X_test,y_train,y_test=TTS(X,y,test_size=0.4)

model.fit(X_train,y_train)
```

```python
#GRADIENT BOOSTING

from sklearn.ensemble import GradientBoostingRegressor

model2=GradientBoostingRegressor()

model2.fit(X_train,y_train)
```

```python
from matplotlib import pyplot as plt

y_pred2=model2.predict(X_test)
y_pred=model.predict(X_test)
plt.scatter(y_pred,y_test)
plt.figure()
plt.scatter(y_pred2,y_test,color='r',marker='x')
plt.show()
```

```python
from sklearn import metrics

metrics.r2_score(y_test,y_pred)
```

```python
metrics.r2_score(y_test,y_pred2)
```
