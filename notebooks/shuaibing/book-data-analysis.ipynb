> There are some errors in the original data set. I modified and re-uploaded the file, which is named ```books(error_correction).csv```,The file encoding is utf-8.

````python
data = pd.read_csv("/kaggle/input/bookserror-correction/books(error_correction).csv",encoding="utf8")
data.head()
````

- 首先，我们来看下所有书籍的评分分布情况

````python
plt.figure(figsize=(16,6))
sns.distplot(data.average_rating,rug_kws={"color":"green"},rug=True)
````

>  很显然，书籍评分在4.0附近，大体成正态分布

````python
# 我们看一下所有书籍语言分布情况
plt.figure(figsize=(16,6))
sns.countplot(data.language_code)
````

> 书籍语言以英文为主，其他语言占很小的比率

````python
plt.figure(figsize=(16,6))
sns.boxplot(data["# num_pages"])
````

> 所有书籍页码分布状况如上图所示，大多数在300-800之间

````python
# 我们可以看下评分数量与评论数量是否有关联
plt.figure(figsize=(16,6))
sns.pointplot(x="ratings_count",y="text_reviews_count",data=data)
plt.xticks(range(0,7000,700),(600000*i for i in range(10)))
````

> 我们观察到开始的时候，随着评分数量的递增，评论数量也呈现出缓慢增加的迹象，但随着评分数量越来越大，评论数量增加也变得急剧起来；到某一个节点的时候，评论数量以指数形式爆炸式的增长。

````python
# 多变量分析,热力图显示变量之间的关系
corrArray = data.corr()
plt.figure(figsize=(6,6))
sns.heatmap(corrArray)
````

> 上面热力图展现了各个变量间的相关性。颜色越浅，代表着两个变量的相关性越大。在图中，我们可以看到，```ratings_count```与```text_reviews_count```具有较强的相关性，这也证实了之前对两个变量的分析。当然，我们也看到```average_rating```与```# num_pages```之间也存在着一定的相关性，下面我们可以进一步证实下。

````python
# 我们可以看下评分与书本页数是否有关联
plt.figure(figsize=(16,6))
sns.pointplot(x="average_rating",y="# num_pages",data=data)
````

> ```average_rating```与```# num_pages```之间的相关性在中间一段表现得比较稳定，但是评分较低或较高时，两个变量之间的关系波动明显

````python
plt.figure(figsize=(16,16))
sns.pairplot(data)
````
