```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../input/used-car-dataset-ford-and-mercedes/cclass.csv')
df2 = pd.read_csv('../input/used-car-dataset-ford-and-mercedes/focus.csv')
```

```python
df.head()
```

```python
df = pd.get_dummies(df)
```

```python
df.head()
```

```python
features = ['year', 'mileage', 'engineSize']
```

```python
target = 'price' 
```

```python
for i in features:
    plt.figure(figsize = (12,5))
    plt.title(f"{i} vs {target}")
    sns.lineplot(x=df[i], y=df[target], data=df, color='b')
    sns.scatterplot(x=df[i], y=df[target], data = df, color = 'r')
    plt.show()
```

```python
df2.head()
```

```python
df.isnull().sum()
```

```python
df_copy = df.copy()
```

```python
y = df_copy['price']
df_copy.drop(df.columns[[1, 4]], axis = 1, inplace = True)
X = df_copy
```

```python
y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 25)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg = PolynomialFeatures(degree = 1)
X_new = poly_reg.fit_transform(X_train)
regressor = LinearRegression()
regressor.fit(X_new, y_train)
```

```python
y_pred = regressor.predict(poly_reg.transform(X_test))
```

```python
data = list(a for a in range(len(y_pred)))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
plt.title('Number of test data VS y_pred and y_test')
plt.plot(data, y_pred)
plt.plot(data, y_test)
```
