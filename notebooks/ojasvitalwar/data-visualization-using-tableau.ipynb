The Dataset Used Here Provides Data of Aiport Traffic Numbers from Various Countries After COVID-19 hit. The Airport traffic is represented as a Percentage of the Baseline, Baseline being February 1-15. Prior to the lockdowns and Social Distancing Measures.

```python
#This Exploratory Data Analysis is done in Tableau
from IPython.display import IFrame
IFrame('https://public.tableau.com/app/profile/ojasvi.talwar/viz/EDA-Covid19ImpactonAirportTraffic-1/Sheet3', width=930, height=500)
```

On First Glance From the Above Visualization we can observe the following.
* Percentage of air traffic at airports on an average did see a hit in the initial two months once the pandemic started. 
* There is variation of magnitude but the trend for the first few months all show a dip in airport traffic as is expected due to the social distancing measures as well as the revised travel guidelines 
* In canada, chile and USA we have seen a bounce back of sorts of the Airport Traffic after the initial Dip. It is still not at the Baseline Levels of Februaruy but have maintained an Upward-Trend
* Australia on the other hand has shown a dip again in the Airport Traffic Post Julyand is at a significantly low level.
