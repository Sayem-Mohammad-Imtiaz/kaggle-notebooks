#  Data visualiation with all common plots.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv("/kaggle/input/iris-flower-dataset/IRIS.csv")
iris.head()
```

# Scatter plot

```python
plt.scatter(iris['sepal_length'],iris['sepal_width'])
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.title('Scatter plot on Iris dataset')
```

# Scatter plot using seaborn

```python
#using seaborn
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue="species", size=6) \
   .map(plt.scatter, "sepal_length", "sepal_width") \
   .add_legend()
plt.show()
```

# Pair plot

```python
sns.set_style("whitegrid");
sns.pairplot(iris, hue="species", size=2);
plt.show()
```

# Box plot

```python
plt.figure(figsize=(20,5))
sns.boxplot(x='sepal_length', y="sepal_width",data=iris,color="orange")
```

# Violin plots

```python
plt.figure(figsize=(20,5))
sns.violinplot(x='sepal_length', y="sepal_width",data=iris, color="orange")
```

# Distribution plot

```python
plt.figure(figsize=(14,5))
sns.distplot(iris['sepal_length'],color="green")
```

# Joint plot

```python
#plt.figure(figsize=(6,5))
sns.jointplot(x ='sepal_length', y ='sepal_width', data = iris, color="red")
```

```python
sns.jointplot(x ='sepal_length', y ='sepal_width', data = iris, kind="kde",color="red")
```

# bar plot with comparision

```python
import numpy as np
import matplotlib.pyplot as plt
data = [[30, 25, 50, 20],
[40, 23, 51, 17],
[35, 22, 45, 19]]
X = np.arange(4)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(X + 0.00, data[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, data[1], color = 'g', width = 0.25)
ax.bar(X + 0.50, data[2], color = 'r', width = 0.25)
ax.legend(labels=['CS', 'IT','ECE'])
plt.show()
```

# Line plot

**simple line plot**

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(1,11)
y = np.random.random(10)
plt.plot(x,y)
plt.show()
```

**different types**

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (12,4))
x = np.arange(1,11)
y1 = np.random.random(10)
y2 = np.random.random(10)
y3 = np.random.random(10)
ax1.plot(x, y1)
ax1.set_title('Plain Line plot')
ax2.plot(x, y2, marker = 'o')
ax2.set_title('Line plot with markers')
ax3.plot(x, y3, marker = 'o', linestyle = ':')
ax3.set_title('Line plot with markers & linestyle')
plt.show()
```

**line plot with color**

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize = (12,4))
x = np.arange(1,11)
y1 = np.random.random(10)
y2 = np.random.random(10)
y3 = np.random.random(10)
ax1.plot(x, y1, marker = '*')
ax1.plot(x, y2, marker = '*')
ax1.set_title('Double Line plot')
ax2.plot(x, y1, marker = '*', linestyle = ':')
ax2.plot(x, y2, marker = '*', linestyle = ':')
ax2.plot(x, y3, marker = '*', linestyle = ':')
ax2.set_title('2 or more Lines')
ax3.plot(x, y1, marker = '*', label = 'line1')
ax3.plot(x, y2, marker = '*' , label = 'line2')
ax3.plot(x, y3, marker = '*', linestyle = ':', label = 'line3')
ax3.set_title('Multiple lines with legends')
plt.legend()
plt.show()
```
