```python
A = np.random.uniform(5, 2.5, (10, 5))
print(np.mean(A[5,:]))
print(np.median(A[:,3]))
print(np.sum(A, axis=1))
print(np.sum(A, axis=0))
print(np.sum(A))
print(np.linalg.det(A[:3,:3]))
y=np.sum(A, axis=1)
z=np.mean(y)
print(z)
d=y[(y<z)]
print(d)
print(d.size)
print(A[A[:, 0]>A[:,4], :])
print(A.shape)
a=np.sum(A, axis=1)
b=np.mean(a)
print(b)
c=a[(a>b)]
print(c)
rank = np.linalg.matrix_rank(A)
print(rank)
```
