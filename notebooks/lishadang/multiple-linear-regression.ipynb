1. Importing the libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

2. Importing the dataset:

```python
df = pd.read_csv("../input/car-price-prediction/CarPrice_Assignment.csv")
df.head(5)
```

```python
unique_value = df["CarName"].nunique()
unique_value
```

3.    Y: dependent variable vector
        #In the first run Xâ€™s type is object due to the different types of 
        independent variables.State column contains categorical variables

```python
x = np.array(df.iloc[:,21:25].values)
y = np.array(df.iloc[:,-1].values)
```

4. Splitting the dataset into the Training and Test dataset
#train_set_split: Split arrays or matrices into random train and #test subsets. %30 of the dataset to the test set

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=16)
```

5. Fit multiple Linear Regression model to our Train set

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

6. Predicting the Test set results:

```python

m = model.coef_
c = model.intercept_
y_predict = sum(m*x_test+c)
y_predict = model.predict(x_test)
y_test_predict = model.predict(x_test)
```

```python
y_predict,y_test_predict
```

```python
plt.scatter(y_test,y_predict,color='red')
plt.plot(y_predict,y_test_predict,color = 'green')
plt.show()
```
