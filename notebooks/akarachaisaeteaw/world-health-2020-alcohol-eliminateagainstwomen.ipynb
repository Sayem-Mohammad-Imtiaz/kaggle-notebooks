```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import glob
import seaborn as sns
import matplotlib.pyplot as plt

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
## using loop for show name of file
for data_file in glob.glob('/kaggle/input/who-worldhealth-statistics-2020-complete/*'):
    df_name = 'df_' + os.path.basename(data_file).split('.')[0]
    print(df_name)
    globals()[df_name] = pd.read_csv(data_file)
```

```python
## Plot count chart between Period and Number Violence Against Women
result_1 = df_eliminateViolenceAgainstWomen.sort_values(by='Period')
fig_dims = (9, 6)
fig, ax = plt.subplots(figsize=fig_dims)
ax = sns.countplot(data=result_1, y='Period')
ax.set_xlabel('Number Violence Against Women',fontsize=12);
ax.set_ylabel('Period',fontsize=12);
```

```python
## Check number of Age
result2 = df_eliminateViolenceAgainstWomen['Dim2'].value_counts().iloc[:10]
result2
```

```python
## Plot Top 10 Range of Age in Pie Chart
labels = ['15-49 years', '25-29 years', '35-39 years', '40-44 years', '45-49 years', '30-34 years',
          '20-24 years', '15-19 years', '18-49 years', '15-24 years'] 
fig_dims = (8, 8)
fig, ax2 = plt.subplots(figsize=fig_dims)
ax2 = plt.pie(result2, autopct='%1.1f%%',
             shadow=True, startangle=90)
plt.legend(labels=labels, loc="center right", title="Range of Age")
plt.title("Top 10 Range of Age in Eliminated Violence Against Women", fontsize=20)
```
