Lets draw CDF (cumulative distribution function) curves

```python
data_survived_more = data[data['surv_st']==1]
data_survived_less = data[data['surv_st']==2]
```

```python
counts , bin_edges = np.histogram(data_survived_more['Age'], bins=10, density=True)
pdf = counts / (sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.show()

counts , bin_edges = np.histogram(data_survived_less['Age'], bins=10, density=True)
pdf = counts / (sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.show()
```

```python
counts , bin_edges = np.histogram(data_survived_more['axi_node'], bins=10, density=True)
pdf = counts / (sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.show()

counts , bin_edges = np.histogram(data_survived_less['axi_node'], bins=10, density=True)
pdf = counts / (sum(counts))
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.show()
```

**Summarising Observations:**
1. Lesses axillary nodes more changes of longer sunvival.
2. Chances of having more axilliary nodes increases after the age of 35(roughly)
3. Younger patients with large number of axilliary nodes have better change of survival as compared with older patients having large axilliary nodes
4. Number of patients operated each year decreasing with year.
