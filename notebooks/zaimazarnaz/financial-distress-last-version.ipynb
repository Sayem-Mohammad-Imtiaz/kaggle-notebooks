```python
ex = np.zeros([32,14, 18,24,23,4])


for i in range(0,330):
    x = company[total_dataframe['Company'][i]]
    y = time[total_dataframe['Time'][i]]
    z = pc1[total_dataframe['principal_component1'][i]]
    s = pc2[total_dataframe['principal_component2'][i]]
    t = pc3[total_dataframe['principal_component3'][i]]
    u = pc4[total_dataframe['principal_component4'][i]]
    #print(str(x)+" "+str(y)+" "+str(z))
    info = total_dataframe['financial_distress'][i]
    
    #print(info)
    ex[x][y][z][s][t][u] = info


for i in range(0,1000):
    a = random.randint(0,31)
    b = random.randint(0,13)
    c = random.randint(0,17)
    d = random.randint(0,23)
    e = random.randint(0,22)
    f = random.randint(0,3)
    if(i == 1):
        print('Co-ordinates:')
        print(a)
        print(b)
        print(c)
        print(d)
        print(e)
        print(f)
    ex[a][b][c][d][e][f] = 1

print(ex)
```

```python

ar = ex.flatten()
print(ar.shape)
y= np.unique(ar)
print(len(y))
```

```python
from tensorly.decomposition import non_negative_parafac
from tensorly.metrics import regression

tensor = tensorly.tensor(ex)
print(tensor.shape)
factors = non_negative_parafac(tensor,rank = 4, verbose = 5)
print(factors)

```

```python
print(len(factors))

print(factors[0].shape)
print(factors[1][0].shape)
print(factors[1][1].shape)
print(factors[1][2].shape)
print(factors[1][3].shape)
print(factors[1][4].shape)
print(factors[1][5].shape)
print(factors[0])
print(factors[1][0])
print(factors[1][1])
```

```python
regenerated_tensor = tensorly.kruskal_to_tensor(factors)
print(regenerated_tensor.shape)
print(regenerated_tensor[5][12][6][10][8][0])
```

```python
#Error measurement
rmse = tensorly.metrics.regression.RMSE(tensor, regenerated_tensor, axis=None)
print("Root Mean Square error = "+str(rmse))
print("Mean squared error = " + str(tf.keras.losses.MSE(tensor,regenerated_tensor)))
```
