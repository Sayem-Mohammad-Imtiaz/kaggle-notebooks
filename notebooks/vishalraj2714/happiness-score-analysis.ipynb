```python
#Let's concatenate all the data sets in to one
newdf= pd.concat([df2015,df2016,df2017,df2018,df2019])
```

```python
newdf
```

```python
pivotdfhappiness = pd.pivot_table(newdf, index= 'Country', columns='Year', values="Happiness_Score")
pivotdfhappiness.fillna(0,inplace=True)
pivotdfhappiness['Average'] = pivotdfhappiness.mean(axis = 1, skipna = True) 
pivotdfhappiness = pivotdfhappiness.sort_values(ascending=False,by='Average')
pivotdfhappiness = pivotdfhappiness.iloc[:35]
pivotdfhappiness = pivotdfhappiness.drop('Average',axis=1)
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(16,10))
pivotdfhappiness.plot(kind='bar',stacked=True,lw = 1,ls=":",fontsize = 10)
plt.legend(loc='upper right',bbox_to_anchor=(1.3, 1),fancybox=True,title='Years',ncol=1)
plt.xticks(rotation = 90)
plt.ylabel('Happiness')
plt.title('Happiness score Trend for 5 Years',fontsize = 10)
plt.tight_layout()
plt.show()
```

```python
#GDP_Per_Capita impact on Happiness score trend for all the 5 Years

pivotdfGDP_per_Capita = pd.pivot_table(newdf, index= 'Country', columns='Year', values="GDP_per_Capita")
pivotdfGDP_per_Capita.fillna(0,inplace=True)
pivotdfGDP_per_Capita['Average'] = pivotdfGDP_per_Capita.mean(axis = 1, skipna = True) 
pivotdfGDP_per_Capita = pivotdfGDP_per_Capita.sort_values(ascending=False,by='Average')
pivotdfGDP_per_Capita = pivotdfGDP_per_Capita.iloc[:35]
pivotdfGDP_per_Capita = pivotdfGDP_per_Capita.drop('Average',axis=1)
plt.style.use('ggplot')
plt.figure(figsize=(10,10))
pivotdfGDP_per_Capita.plot(kind='bar',stacked=True,lw = 1,ls=":",fontsize = 10)
plt.legend(loc='upper right',bbox_to_anchor=(1.3, 1),fancybox=True,title='Years',ncol=1)
plt.xticks(rotation = 90)
plt.ylabel('GDP_per_Capita')
plt.title('GDP_per_Capita score Trend for 5 Years',fontsize = 10)
plt.tight_layout()
plt.show()
```

```python
#Freedom impact on Happiness score trend for all the 5 Years

pivotdffreedom = pd.pivot_table(newdf, index= 'Country', columns='Year', values="Freedom")
pivotdffreedom.fillna(0,inplace=True)
pivotdffreedom['Average'] = pivotdffreedom.mean(axis = 1, skipna = True) 
pivotdffreedom = pivotdffreedom.sort_values(ascending=False,by='Average')
pivotdffreedom = pivotdffreedom.iloc[:35]
pivotdffreedom = pivotdffreedom.drop('Average',axis=1)
plt.style.use('ggplot')
plt.figure(figsize=(10,10))
pivotdffreedom.plot(kind='bar',stacked=True,lw = 1,ls=":",fontsize = 10)
plt.legend(loc='upper right',bbox_to_anchor=(1.3, 1),fancybox=True,title='Years',ncol=1)
plt.xticks(rotation = 90)
plt.ylabel('Freedom')
plt.title('Freedom score Trend for 5 Years',fontsize = 10)
plt.tight_layout()
plt.show()
```
