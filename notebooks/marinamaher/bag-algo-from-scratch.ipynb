# Import

```python
import numpy as np 
import pandas as pd
from scipy.stats import norm
from sklearn.preprocessing import LabelEncoder
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split 
from sklearn import metrics 
```

# Data

```python
df = pd.read_csv("../input/cardiovascular-disease-dataset/cardio_train.csv",sep=";")
df.drop('id', inplace=True, axis=1)
X, y = df.iloc[:,:df.shape[1]-1] , pd.DataFrame(df.iloc[:,df.shape[1]-1])
```

# Bagging

```python
def bag(n_bins,X):
    divide = int(X.shape[0]/n_bins)
    i=0
    l=[]

    for j in range(n_bins):
        l.append(i)
        i = i+divide

    if(l[len(l)-1] < X.shape[0]):
        l.append(X.shape[0])
        
    dictt={}
    for i in range(len(l)-1):
        dictt[(l[i],l[i+1])]=False

    return dictt

```

# Train & test

```python
n_bins=3
d = bag(n_bins,X)
acc=[]

for key in d:
    X_test = X[int(key[0]): int(key[1])]
    y_test = y[int(key[0]): int(key[1])]
    d[key]=True
    X_train= pd.DataFrame()
    y_train=pd.DataFrame()
    for key_train in d :
        if ( d[key_train] == False):
            X_train = X_train.append( X.iloc[key_train[0] : key_train[1],:] )
            y_train = y_train.append( y.iloc[key_train[0] : key_train[1]] )
        
    clf = DecisionTreeClassifier()
    clf = clf.fit(X_train,y_train)
    y_pred = clf.predict(X_test)
    print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
    acc.append(metrics.accuracy_score(y_test, y_pred))
    d[key]= False
print("Final accuracy", np.mean(acc))
```
