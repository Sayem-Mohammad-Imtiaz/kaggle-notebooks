**loading the data into the df **

```python
df=pd.read_csv('../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
df.head()
```

**checking how many fiction and non fiction books are present **

```python
sns.countplot(x='Genre',data=df)
plt.show()
```

```python
most_frequent_authors=df.Author.value_counts()
```

```python
sns.countplot(x='Author',data=df)
```

```python
top_5_author=(df.groupby(by='Author',as_index=False)
              .agg(({'Price':'sum'}))
              .sort_values('Price', ascending=False))[:5]
print(top_5_author)
```

```python
top_5_user_rating=(df.groupby(by='User Rating',as_index=False)
              .agg(({'User Rating':'sum'}))
              .sort_values('User Rating', ascending=False))[:5]
print(top_5_user_rating)
```

```python
minimum_rating = (
    df.groupby(['Author','Name', 'Price'], as_index=False)
    .agg({'User Rating':'sum'})
    .sort_values('User Rating', ascending=True))[:5]
minimum_rating
```

```python
maximum_rating = (
    df.groupby(['Author','Name', 'Price'], as_index=False)
    .agg({'User Rating':'max'})
    .sort_values('User Rating', ascending=False))[:5]
maximum_rating
```

```python
max_Price= (
    df.groupby(['Author','Name', 'Price'], as_index=False)
    .agg({'Price':'max'})
    .sort_values('Price', ascending=False))[:5]
max_Price
```
