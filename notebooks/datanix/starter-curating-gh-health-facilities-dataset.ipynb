 ## <a id="load">1. Loading and Inspecting the dataset </a>

```python
import pandas as pd

df_health_facilities = pd.read_csv("../input/health-facilities-gh.csv")
df_health_facilities.head(5)
```

```python
# Filter out facilities of interest Type=='Others'
df_hf_others = df_health_facilities[df_health_facilities['Type']=='Others'].copy()
df_hf_others
```

## <a id="explore"> 2. Dealing with Uncategorised Facilities </a>
If the name contains `"Feeding|Nutrition|Rehabilitation"` let's call it `Feeding Center` instead of `Others`.  
Hisham's Notebook has more details.


```python
df_hf_others.loc[df_hf_others.FacilityName.str.contains("Feeding|Nutrition|Rehabilitation"),'Type'] = "Feeding Center"

df_hf_others
#Hurray !!!!
```
