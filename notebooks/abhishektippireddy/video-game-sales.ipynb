```python
gamesales.Platform.unique()
```

```python
def platform_to_company(row):
    if row.Platform in ['PS','PS2','PS3','PS4','PSP','PSV']:
        row.Platform='Sony'
    elif row.Platform in ['X360','XOne','XB']:
        row.Platform='Microsoft'
    elif row.Platform in ['DS','Wii','GBA','NES','WiiU','N64','SNES','GB','3DS','GC']:
        row.Platform='Nintendo'
    elif row.Platform=='PC':
        row.Platform='computer'
    else:
        row.Platform='others'
    return row
```

```python
gamesales.apply(platform_to_company,axis='columns').groupby('Platform').Global_Sales.count().sort_values().plot(kind='pie',figsize=(8,8))
plt.title('Number of Games on a Companies Platform')
plt.ylabel('')
```

# Yearly Game Sales

```python
gamesales.groupby('Year').Name.count().sort_values(ascending=False).head(7).plot(kind='bar',figsize=(10,7))
plt.title('Game sales by year')
plt.ylabel('Sales (in millions)')
```

> **2009 has the highest number of game titles**

```python
gamesales.groupby('Year').Global_Sales.sum().plot(kind='line',figsize=(12,8),marker='o',grid=True)
plt.title('Yearly sales (Global)')
plt.ylabel('Sales (in millions)')
```

> **2008 has the highest number of sales**

> **all the years with most number of game title have the most number of sales**

# Which Genre has the most number of Games and sales

```python
gamesales.Genre.value_counts().sort_values().plot(kind='barh',figsize=(10,6))
plt.title('Total number of games by genre')
plt.xlabel('Number of units(count)')
plt.ylabel('Genre')
```

> **Action games are the most number of titles followed by sports**

```python
gamesales.groupby('Genre').Global_Sales.sum().sort_values().plot(kind='barh',figsize=(10,6))
plt.title('Global sales by Genre')
plt.xlabel('Sales (in millions)')
```

> **With action games being the most number of games released they have the highest number of sales but it doesnt look like the case with all genre's of games**

```python
gamesales.groupby('Genre').NA_Sales.sum().sort_values().plot(kind='barh',figsize=(10,6))
plt.title('Naorth American Sales by Genre')
plt.xlabel('Sales (in millions)')
```

> **North America is playing similar games that the entire world is playing**

```python
gamesales.groupby('Genre').EU_Sales.sum().sort_values().plot(kind='barh',figsize=(10,6))
plt.title('Europe sales by Genre')
plt.xlabel('Sales (in millions)')
```

> **Europe is also intrested in playing Action Games followed by sports games**

```python
gamesales.groupby('Genre').JP_Sales.sum().sort_values().plot(kind='barh',figsize=(10,6))
plt.title('Japan sales by Genre')
plt.xlabel('Sales (in millions)')
```

> **Looks like Japan is more into Role-Playing Games unlike the other two regions**

```python
gamesales.groupby('Genre').Other_Sales.sum().sort_values().plot(kind='barh',figsize=(10,6))
plt.title('Rest of the World sales by Genre')
plt.xlabel('Sales (in millions)')
```

# Who are the top publishers in the past decade

```python
gamesales[gamesales.Year>2010].groupby('Publisher').Global_Sales.sum().sort_values().tail(10).plot(kind='barh',figsize=(10,7))
plt.title('Most sales in last Decade (World)')
plt.xlabel('Sales (in millions)')
```

> **Electronic Arts has the most number of sales in the entire world for the past decade**

```python
gamesales[gamesales.Year>2010].groupby('Publisher').NA_Sales.sum().sort_values().tail(10).plot(kind='barh',figsize=(10,7))
plt.title('Most sales in last Decade (North America)')
plt.xlabel('Sales (in millions)')
```

> **The competition is very close in North America between Activision and Electronic Arts**

```python
gamesales[gamesales.Year>2010].groupby('Publisher').EU_Sales.sum().sort_values().tail(10).plot(kind='barh',figsize=(10,7))
plt.title('Most sales in last Decade (Europe)')
plt.xlabel('Sales (in millions)')
```

> **There is no surprise here with Euopeans being the biggest fans of football and Electronic Arts main title being Fifa it has the highest number of sales in Europe**

```python
gamesales[gamesales.Year>2010].groupby('Publisher').JP_Sales.sum().sort_values().tail(10).plot(kind='barh',figsize=(10,7))
plt.title('Most sales in last Decade (Japan)')
plt.xlabel('Sales (in millions)')
```

> **With Nintendo having the most number of sales in Japan none of the other publishers who performed very well in other regions are in the top 10**
