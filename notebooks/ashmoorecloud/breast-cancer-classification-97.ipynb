```python
input_cols = list(df.columns)[2:-1]
target_col = 'diagnosis'
```

```python
targets = df[target_col].copy()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(df[input_cols])
```

```python
X = scaler.transform(df[input_cols])
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, targets, test_size=0.25, random_state=42)
X_valid, X_test, y_valid, y_test = train_test_split(X_test, y_test, test_size=0.3, random_state=42)
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='liblinear')
```

```python
y_train
```

```python
model.fit(X_train, y_train)
```

```python
train_preds = model.predict(X_train)
```

```python
train_preds
```

```python
train_probs = model.predict_proba(X_train)
```

```python
model.classes_
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_train, train_preds)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_train, train_preds, normalize='true')
```

```python
def predict_and_plot(inputs, targets, name=''):
    preds = model.predict(inputs)
    
    accuracy = accuracy_score(targets, preds)
    print("Accuracy: {:.2f}%".format(accuracy * 100))
    
    cf = confusion_matrix(targets, preds, normalize='true')
    plt.figure()
    sns.heatmap(cf, annot=True)
    plt.xlabel('Prediction')
    plt.ylabel('Target')
    plt.title('{} Confusion Matrix'.format(name));
    
    return preds
```

```python
train_preds = predict_and_plot(X_train, y_train, 'Training')
```

```python
val_preds = predict_and_plot(X_valid, y_valid, 'Validatiaon')
```

```python
val_preds = predict_and_plot(X_test, y_test, 'Test')
```
