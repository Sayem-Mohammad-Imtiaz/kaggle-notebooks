```python
for col in data.columns:
    print("{0} unique vals in {1}".format(data[col].drop_duplicates().count(), col))
```

```python
for name in data['name'].drop_duplicates():
    s = sum(data.tweets.str.contains(name))
    if (s > 0):
        print("{0} has sum {1}".format(name,s))
```
