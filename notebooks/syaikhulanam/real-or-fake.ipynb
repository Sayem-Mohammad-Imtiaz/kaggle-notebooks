# **Membuang Stopwords**

Stopwords adalah kata-kata bahasa Inggris yang tidak menambah banyak arti pada kalimat. Mereka dapat dengan aman diabaikan tanpa mengorbankan arti kalimat. 

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
import re
import string
```

```python
def cleaning_text(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub("\\W"," ",text) # remove special chars
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    
    return text
```

```python
data_news['text']=data_news['text'].apply(cleaning_text)
```

```python
data_news
```

# **Split Data**

```python
x_train,x_test,y_train,y_test=train_test_split(data_news['text'],data_news['label'],test_size=0.25)
```

**Mengkonfersi text ke dalam sebuah vektor**

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(min_df=0,max_df=1,ngram_range=(1,2))

xv_train=cv.fit_transform(x_train)

xv_test=cv.transform(x_test)
```

# **Penerapan Decision Tree Klasifikasi**

```python
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```python
DTC = tree.DecisionTreeClassifier()
DTC.fit(xv_train, y_train)
```

```python
DTC.score(xv_test,y_test)
```
