## Preprocess

```python
df
```

```python
#split the categorical and binary value
#but we check which one is binary columns
for i in df.columns:
    if len(np.unique(df[i]))<=2:
        display(i,np.unique(df[i]))
```

```python
#Drop the output Columns
label = df.pop('output')

# Binary Columns
bin_cols = ['sex','fbs','exng']

# Numerical Columns
num_cols = [x for x in df.columns if x not in bin_cols]
```

```python
# Splitting the train and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df,label, test_size=0.2,random_state=42)
```

```python
#Preprocessing non-binary columns
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df[num_cols] = sc.fit_transform(df[num_cols])
```

```python
df
```

## Machine Learning Model

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
knn_clf = KNeighborsClassifier()
knn_clf.fit(X_train,y_train)
```

```python
knn_pred = knn_clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
display(
    confusion_matrix(y_test,knn_pred),
    accuracy_score(y_test,knn_pred)
)
```

```python
# XGBoost
from xgboost import XGBClassifier
xgb_clf = XGBClassifier()
xgb_clf.fit(X_train,y_train)
```

```python
xgb_pred = xgb_clf.predict(X_test)
```

```python
display(
    confusion_matrix(y_test,xgb_pred),
    accuracy_score(y_test,xgb_pred)
)
```
