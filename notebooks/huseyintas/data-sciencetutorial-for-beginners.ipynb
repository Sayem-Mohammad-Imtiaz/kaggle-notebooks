## Python Data Science  Toolbox

```python
# USER DEFINED FUNCTION
def tuble_ex():
    t = (1,2,3)
    return t
a,b,c = tuble_ex()
print(a,b,c)
```

```python
lis = [1,2,3]
a,b,c = lis
print(a,b,c)
```

```python
# SCOPE
x = 5
def f():
    y = 2 * x # local bir x eğişkeni olmadıgı için global x üzerinden işlem yapar
    return y
print(f())
```

```python
x = 5
def f():
    x = 3
    y = 2 * x # local x degeri oldugu için local x degeri üzerinde işlem yaptı
    return y
print(f())
```

```python
import builtins
dir(builtins)
```

```python
# NESTED Functions
def square():
    def add():
        x = 2
        y = 3
        z = x+y
        return z
    return add()**2
print(square())
```

```python
# Default and Flexible Arguments
def f(a,b=1,c=2):
    y = a+b+c
    return y
print(f(5))
print(f(5,4,3))
```

```python
def f(*args):
    for i in args:
        print(i)
f(1,2,3,4)


def ff(**kwargs):
    for key,value in kwargs.items():
        print(key, ":", value)
ff(country ="france", capital="paris", population = 15005587)
```

```python
# Lambda Functions
square = lambda x: x**2
print(square(4))
```

```python
# Anonymous Functions
liste =[1,2,3,4,5]
y = map(lambda x: x**2,liste)
print(list(y))
```

```python
# ITERATORS
name = "ronaldo"
it = iter(name)
print(next(it))
print(next(it))
print(*it)
```

```python
# zip() methodu
list1 = [1,2,3,4,5]
list2 = [6,7,8,9]
z = zip(list1,list2)
print(z)
z_list = list(z)
print(z_list)
```

```python
un_zip = zip(*z_list)
un_list1,un_list2 = list(un_zip)
print(un_list1)
print(un_list2)
print(type(un_list1))
```

```python
# LIST COMPREHENSİON
num1 = [1,2,3]
num2 = [i+1 for i in num1]
print(num2)
```

```python
num1 = [5,10,15]
num2 = []
for i in num1:
    if i == 10:
        i = i**2
        num2.append(i)
    elif i<7:
        i = i-5
        num2.append(i)
    else:
        i = i+5
        num2.append(i)
print(num2)
```

```python
num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)
```
