- Создайте равномерную сетку из 20 точек на отрезке $\left[0; \frac{\pi}{2}\right]$.
- Проверьте для каждой точки выполнение тригонометрического тождества $\cos\alpha = \sin\left(\frac{\pi}{2}-\alpha\right)$. Для этого: а) создайте массив $\cos\alpha$; б) создайте массив $\sin\left(\frac{\pi}{2}-\alpha\right)$; в) создайте логическую функцию **close(x, y, eps)**, которая возвращает **True**, если $|x-y|<\varepsilon$, и **False** в противном случае; г) примените данную функцию к двум массивам, созданным выше; д) убедитесь, что все значения в полученном массиве равны **True**.

```python
m20=np.linspace(0,np.pi,20)
m20
```

```python
cos_m20=np.cos(m20)
cos_m20
```

```python
sin_m20=np.sin(np.pi/2-m20)
sin_m20
```

```python
cos_m20-sin_m20
```

```python
np.abs(cos_m20-sin_m20)<1e-6
```

```python
def close (x,y,eps):
    if np.abs(x-y)<eps:
        return True
    else:
        return False
```

```python
close_v=np.vectorize(close)
close_v(cos_m20,sin_m20,1e-6)
```

```python
df = np.loadtxt('../input/cardio_train.csv', delimiter=';', skiprows=1)
print(df[:5, :])
```
