```python
correlation = df.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, vmax=1, square=True,annot=True)

plt.title('Correlation between different fearures')
```

Based on the correlation chart above, we can see the following relationships into two different types:

High Positive Relationships:

 - Participants (Course Content Accessed) vs Audited (>50% Course Content Accessed)
 - % Female vs Median Age
 - Certified vs Audited (>50% Course Content Accessed)
 - % Certified of > 50% Course Content Accessed vs % Grade Higher Than Zero
 - Certified vs Total Course Hours (Thousands)
 - % Certified vs % Certified of > 50% Course Content Accessed

High Negative Relationships:

 - %Male vs Median Age
 - Certified vs Year
 - Median Hours for Certification vs Median Age
 - Certified vs % Audited
 - % Audited vs Median Hours for Certification
 - Year vs % Certified of > 50% Course Content Accessed
