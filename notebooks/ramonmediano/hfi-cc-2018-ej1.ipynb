```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../input/the-human-freedom-index/hfi_cc_2018.csv")
df
```

```python
 # Obtén una lista con los atributos cuantitativos y otra lista con los atributos cualitativos
    
lista_cuantitativa = list()
lista_cualitativa = list()

for i in df.columns[4:]:
    df[i] = df[i].replace(np.nan,df[i].mean())  #remplazar los nan por la media de cada columna
    lista_cuantitativa.append(i)
    
for i in df.columns[:4]:
    lista_cualitativa.append(i)
    
```

```python
# Define el rango de cada uno de los atributos (cuantitativos).

lista_rangos = list()
for i in df.columns[4:]:
    lista_rangos.append(np.max(df[i])-np.min(df[i]))
```

```python
# Obtén la moda 

import scipy
from scipy import stats 

lista_moda = list()

for i in df.columns[4:]:
        lista_moda.append(stats.mode(np.array(df[i])))
lista_moda
```

```python
# Obtén el moda de los cualitativos 

import scipy
from scipy import stats 

lista_moda_cualitativos = list()

for i in df.columns[:4]:
        lista_moda_cualitativos.append(stats.mode(np.array(df[i])))
lista_moda_cualitativos
```

```python
#Obtén el promedio 

lista_promedio = list()
for i in df.columns[4:]:
    lista_promedio.append(np.mean(np.array(df[i])))
lista_promedio
```

```python
# Obtener varianza
lista_varianza = list()
for i in df.columns[4:]:
    lista_varianza.append(np.var(np.array(df[i])))
lista_varianza
```

```python
#Obtener desviación

lista_desviacion = list()
for i in df.columns[4:]:
    lista_desviacion.append(np.std(np.array(df[i])))
lista_desviacion
```
