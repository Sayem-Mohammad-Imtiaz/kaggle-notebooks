```python
col = ['age', 'sex', 'cp', 'fbs', 'restecg', 'trestbps',
       'exang', 'oldpeak', 'slope', 'ca', 'thal']
```

```python
plt.style.use('ggplot')
for item in col:
    pd.crosstab(df[item], df.target).plot(kind='bar', figsize=(15, 7))
    plt.title("{} with target".format(str(item)))
    plt.legend(["no", "si"])
    plt.ylabel("Frequency")
plt.show()
```
