```python
# Required Libararies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading data into data frame
df = pd.read_csv('/kaggle/input/boston-housing-dataset/HousingData.csv')
```

```python
# printing head of data frame
df.head()
```

# EDA

```python
# Shape of dataset
df.shape
```

```python
# Structure of Dataset
df.info()
```

```python
# Null Values
df.isnull().sum()
```

```python
# Zeros
df[df==0].sum()
```

# Visualization

```python
# Box Plots
cols = list(df.columns)
for col in cols:
    plt.figure()
    sns.boxplot(y=df[col], color = 'r')
    plt.title(col)
```

# Data Transformation

```python
# Handling null values
df = df.interpolate(method ='linear', limit_direction ='backward', limit = 2)
```

```python
# Dropping null value
df = df.dropna()
```

```python
# Checking Null Values
df.isnull().sum()
```

```python
# shape of data
df.shape
```

```python
# Checking outliers
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
lowerlimit = Q1 - 1.5*IQR
upperlimit = Q3 + 1.5*IQR
```

```python
df_out = df[~((df < lowerlimit) |(df > upperlimit)).any(axis=1)]
print(df_out.shape)
```

# X and y

```python
X = df.drop('MEDV', axis = 1)
y = df.MEDV
```

# Splitting into Train and Test data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

 # Model Builing

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
```

# Fitting Model

```python
linear_reg.fit(X_train,y_train)
```

# Predictions

```python
pred = linear_reg.predict(X_test)
pred
```

# Mean Absolute Error

```python
from sklearn.metrics import mean_absolute_error
print('Mean Squared Error is ',mean_absolute_error(pred,y_test))
```
