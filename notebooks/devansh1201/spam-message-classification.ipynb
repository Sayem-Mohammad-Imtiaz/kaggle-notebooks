```python
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix, classification_report
```

```python
df = pd.read_csv('/kaggle/input/spam-text-message-classification/SPAM text message 20170820 - Data.csv')
df.head()
```

```python
print(df.isnull().sum())
print(df['Category'].unique())
print(df['Category'].value_counts())
```

```python
def fit_metrics(X_train, y_train, X_test, y_test, setNumber):
    print(f"This is for set number: {setNumber} \n")
    text_clf = Pipeline([('tfidf', TfidfVectorizer()), ('clf', LinearSVC())])
    text_clf.fit(X_train, y_train)
    predictions = text_clf.predict(X_test)
    print('Confusion Matrix')
    print(confusion_matrix(y_test, predictions))
    print(classification_report(y_test, predictions))
    print()
```

```python
# keeping a random_state value to get same results for tests.
splits = StratifiedShuffleSplit(n_splits=5, test_size=0.3, random_state=42)
X = df['Message']
y = df['Category']
splits.get_n_splits(X, y)
```

```python
i = 0
for train_index, test_index in splits.split(X,y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    fit_metrics(X_train, y_train, X_test, y_test, i)
    i = i + 1
```
