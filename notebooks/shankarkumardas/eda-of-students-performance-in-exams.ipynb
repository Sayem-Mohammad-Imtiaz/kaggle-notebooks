```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import plotly.express as px
```

```python
student=pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
student.head()
```

```python
student.info()
```

```python
student.describe()
```

```python
gender=student['gender'].value_counts()
gender
```

```python
#Making a bar plot with gender
plt.figure(figsize=(10,5))
plt.bar(list(student['gender'].value_counts().keys()),list(student['gender'].value_counts()),color=["blue","red"])
plt.title("Bar plot of gender")
plt.xlabel("Gender")
plt.ylabel("Member of gender")
plt.grid(True)
plt.show()
```

```python
group=student['race/ethnicity'].value_counts()
group
```

```python
plt.figure(figsize=(10,5))
plt.bar(list(student['race/ethnicity'].value_counts().keys()),list(student['race/ethnicity'].value_counts()),color=['blue','red','orange','green','yellow'])
plt.title("Bar plot of Race/ethnicity")
plt.xlabel("Group")
plt.ylabel("Group Member")
plt.grid(True)
plt.show()
```
