```python
df = pd.read_csv('/kaggle/input/vehicle-dataset-from-cardekho/car data.csv')
```

```python
df.head()
```

```python
df2 = df.drop('Car_Name',axis='columns')
```

```python
data = pd.get_dummies(df2)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
```

```python
data.columns
```

```python
x = data.drop('Selling_Price',axis=1)
y = data['Selling_Price']
```

```python
x
```

```python
train_data, test_data, train_label,test_label = train_test_split(x,y, test_size=0.2)
```

```python
len(train_data)
```

```python
from sklearn import preprocessing
train_data = preprocessing.normalize(train_data)
test_data = preprocessing.normalize(test_data)
```

```python
from tensorflow.keras import models
from tensorflow.keras.layers import *
```

```python
model = models.Sequential()
```

```python
model.add(Dense(512,activation='relu',input_shape=(train_data.shape[1],)))
model.add(Dense(64,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='rmsprop',loss='mae',metrics=['mae'])
```

```python
model.fit(train_data,train_label,epochs=10)
```

```python
test_results = model.evaluate(test_data, test_label, verbose=0)
```
