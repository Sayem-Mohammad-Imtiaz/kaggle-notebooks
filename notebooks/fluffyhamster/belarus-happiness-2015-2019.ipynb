Так как во всех данных есть Belarus, то по ней и будет анализ (Since all data contains Belarus, then will be an analysis for Belarus)

```python
Belarus2015 = data2015.loc[data2015["Country"]=="Belarus"]
Belarus2015 = Belarus2015.drop(["Region", "Standard Error", 'Dystopia Residual'], axis=1)
Belarus2015 = Belarus2015.rename(columns={"Country":"Country or region",
                                          "Happiness Score": "Score", 
                                          "Happiness Rank":"Overall rank", 
                                          "Economy (GDP per Capita)":"GDP per capita",
                                          "Health (Life Expectancy)" : "Healthy life expectancy",
                                          "Trust (Government Corruption)": "Perceptions of corruption",
                                          "Family":"Social support",
                                          "Freedom": "Freedom to make life choices"})
Belarus2015
```

```python
Belarus2016 = data2016.loc[data2016["Country"]=="Belarus"]
Belarus2016 = Belarus2016.drop(["Region", 'Dystopia Residual', 'Lower Confidence Interval', 'Upper Confidence Interval'], axis=1)
Belarus2016 = Belarus2016.rename(columns={"Country":"Country or region",
                                          "Happiness Score": "Score", 
                                          "Happiness Rank":"Overall rank", 
                                          "Economy (GDP per Capita)":"GDP per capita",
                                          "Health (Life Expectancy)" : "Healthy life expectancy",
                                          "Trust (Government Corruption)": "Perceptions of corruption",
                                          "Family":"Social support",
                                          "Freedom": "Freedom to make life choices"})
Belarus2016
```

```python
Belarus2017 = data2017.loc[data2017["Country"]=="Belarus"]
Belarus2017 = Belarus2017.drop(['Whisker.high', 'Whisker.low', 'Dystopia.Residual'], axis=1)
Belarus2017 = Belarus2017.rename(columns={"Country":"Country or region",
                                          "Happiness.Score": "Score", 
                                          "Happiness.Rank":"Overall rank", 
                                          "Economy..GDP.per.Capita.":"GDP per capita",
                                          "Health..Life.Expectancy." : "Healthy life expectancy",
                                          "Trust..Government.Corruption.": "Perceptions of corruption",
                                          "Family":"Social support",
                                          "Freedom": "Freedom to make life choices"})
Belarus2017
```

```python
Belarus2018 = data2018.loc[data2018["Country or region"]=="Belarus"]
Belarus2019 = data2019.loc[data2019["Country or region"]=="Belarus"]
```

```python
belarus = pd.concat([Belarus2015, Belarus2016, Belarus2017, Belarus2018, Belarus2019])
belarus = belarus.rename(index={58:"2015", 60:"2016", 66:"2017", 72:"2018", 80:"2019"})
belarus = belarus.drop('Country or region', axis=1)
belarus
```
