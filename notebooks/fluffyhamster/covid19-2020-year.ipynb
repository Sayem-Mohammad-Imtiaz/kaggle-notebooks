# Plot 

```python
def plot_count(col_name):
    df = pd.DataFrame(data[col_name].value_counts())
    
    def func(pct, allvals):
        absolute = int(pct/100.*np.sum(allvals))
        return "{:.1f}%".format(pct, absolute)
    
    df.plot.pie(y=col_name, labels=None, figsize=(10, 10), title=col_name, autopct=lambda pct: func(pct, df[col_name])).legend(title=col_name, labels=df.index, loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
    plt.show()
```

```python
for col_name in ['current_status', 'sex', 'age_group', 'Race and ethnicity (combined)', 'hosp_yn', 
        'icu_yn', 'death_yn', 'medcond_yn']:
    plot_count(col_name)
```
