# Data

```python
data = pd.read_csv("../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv")
```

```python
data = data.sort_values(by="Year")
data.head()
```

# The most expensive book (Самая дорогая книга)

```python
data.loc[data["Price"] == data["Price"].max()]
```

# Count "Fiction" and "Not Fiction"

```python
data.groupby("Genre").Genre.count()
```

```python
count_year_genre = pd.DataFrame(data.groupby(["Year", "Genre"]).Year.count())
count_year_genre = count_year_genre.rename(columns={"Year" : "count"})
count_year_genre = count_year_genre.reset_index()
count_year_genre
```

```python
import seaborn as sns
plt = sns.lmplot(x = 'Year', y = "count", data=count_year_genre, hue="Genre")
```

В 2014 году было больше фантастики чем в другие годы, список этих книг:

```python
fiction = data.loc[data.Genre == "Fiction"]
fiction.loc[fiction.Year == 2014]
```

# Better Autors

```python
data = data.rename(columns={"User Rating":"User_Rating"})
```

```python
author_rating = pd.DataFrame(data.groupby("Author").User_Rating.mean()).reset_index()
author_rating = author_rating.sort_values(by="User_Rating")
author_rating.loc[author_rating["User_Rating"] == author_rating["User_Rating"].max()]
```

# Better Autor, max books in top

```python
count_books_Author = pd.DataFrame(data.groupby("Author").Author.count())
count_books_Author = count_books_Author.rename(columns={"Author" : "count_books"}).reset_index()
count_books_Author = count_books_Author.sort_values(by="count_books", ascending=False)
count_books_Author.head(1)
```

```python
data.loc[data.Author == "Jeff Kinney"].sort_values(by="Year")
```
