```python
dfdaily = pd.read_csv('../input/corona-virus-report/day_wise.csv')
dfdaily.head()
```

```python
dfdaily['Date'] = pd.to_datetime(dfdaily['Date'])
dfdaily['Month'] = dfdaily['Date'].apply(lambda time: time.month)
dfdaily['Day of Week'] = dfdaily['Date'].apply(lambda time: time.dayofweek)
dfdaily['Year'] = dfdaily['Date'].apply(lambda time: time.year)
```

```python
dfdaily.drop('Date',axis=1,inplace=True)
```
