```python
#taking thousand rows
new_salt_temp = salt_temp[:1000]
```

```python
new_salt_temp = new_salt_temp.dropna(axis=0, how="any")
```

```python
new_salt_temp.isnull().values.any()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(new_salt_temp['Salnty'],new_salt_temp['T_degC'],alpha = 0.7)
plt.xlabel('Salinity')
plt.ylabel('Temp')
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(new_salt_temp)
```

```python
x = scaled_data[:,0]
```

```python
y = scaled_data[:,1]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
lin_reg_model = reg.fit(x_train.reshape(-1, 1),y_train)
```

```python
y_pred = lin_reg_model.predict(x_test.reshape(-1, 1))
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
plt.scatter(x_test,y_pred)
plt.scatter(x_test,y_test,alpha = 0.5)
plt.xlabel('Salinity')
plt.ylabel('Temp')
```
