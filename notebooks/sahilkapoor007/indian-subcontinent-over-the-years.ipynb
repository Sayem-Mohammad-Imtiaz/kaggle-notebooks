```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data_filepath = '../input/world-happiness-report-2021/world-happiness-report.csv'
data = pd.read_csv(data_filepath)
data
```

```python
subcontinent = data[(data['Country name'] == 'India') | (data['Country name'] == 'Pakistan') | (data['Country name'] == 'Bangladesh') | (data['Country name'] == 'Sri Lanka') | (data['Country name'] == 'Nepal') | (data['Country name'] == 'Afghanistan')]
subcontinent
#ind = data[data['Country name'] == 'India']
#pak = data[data['Country name'] == 'Pakistan']
#ban = data[data['Country name'] == 'Bangladesh']
#nep = data[data['Country name'] == 'Nepal']
#sl = data[data['Country name'] == 'Sri Lanka']
#afg = data[data['Country name'] == 'Afghanistan']
```

# Indian subcontinent over the years

```python
plt.figure(figsize=(20,10))
sns.set_style('darkgrid')
sns.lineplot(x='year', y='Life Ladder', data=subcontinent, marker='o', hue='Country name', palette='hls', linewidth=3)
plt.title('Life Ladder vs Year',fontdict= { 'fontsize': 24, 'fontweight':'bold'})
plt.xlabel('Years', fontsize=18)
plt.ylabel('Life Ladder', fontsize=18)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0, prop={'size':15})
```

```python
plt.figure(figsize=(20,10))
sns.set_style('darkgrid')
sns.lineplot(x='year', y='Perceptions of corruption', data=subcontinent, marker='o', hue='Country name', palette='hls', linewidth=3)
plt.title('Perception of corruption over the years',fontdict= { 'fontsize': 24, 'fontweight':'bold'})
plt.xlabel('Years', fontsize=18)
plt.ylabel('Perception of corruption', fontsize=18)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0, prop={'size':15})
```
