```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [10, 10]
```

```python
import math
def roundup(x):
    return 100 + int(math.ceil(x / 100.0)) * 100 
sns.countplot(x='rating',data=netflix)
# Get current axis on current figure
ax = plt.gca()

# ylim max value to be set
y_max = netflix['rating'].value_counts().max() 
ax.set_ylim([0, 3000])

# Iterate through the list of axes' patches
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
            fontsize=12, color='black', ha='center', va='bottom')
plt.title('Rating Counts Bar Chart',size='15')
plt.show()
```

```python
ratings = netflix.groupby(['rating'])['show_id'].count().reset_index(name='count').sort_values(by='count',ascending=False)
ratings.head()
```

```python
plt.figure(figsize=(10,10))
# make barplot and sort bars in descending order
sns.barplot(x='rating', 
            y='count', 
            data=ratings, 
            order=ratings.sort_values('count',ascending = False).rating)
#set axis for count label
ax = plt.gca()
# set labels
plt.xlabel("Rating", size=15)
plt.ylabel("Count", size=15)
# Iterate through the list of axes' patches
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
            fontsize=12, color='black', ha='center', va='bottom')
plt.title("Sort Ratings in Descending Order", size=18)
plt.tight_layout()
plt.show()
```

```python
years = netflix.groupby(['release_year'])['show_id'].count().reset_index(name='count').sort_values(by='count',ascending=False)

fig_dims = (18,7)

fig, ax = plt.subplots(figsize=fig_dims)
sns.pointplot(x='release_year',y='count',data=years.head(19))

plt.title('Release Count by Year',size='20')
plt.xlabel("Release Year")
plt.ylabel("Release Count")
plt.show()
```

```python
#Cast exploration. Need to remove delimiter and expand fields with multiple actors/actresses to complete analysis on cast field. 
actors = netflix.groupby(['cast'])['show_id'].count().reset_index(name='count').sort_values(by='count',ascending=False)
actors.head()
```

```python
#Cast isolation completed using str.split and expand commands. Each record now only has one cast name instead of multiple.
cast2 = netflix.set_index("title").cast.str.split(',', expand=True).stack().reset_index(level=1, drop=True)
print(cast2)
```

```python
cast2.head(100)
```

```python

plt.figure(figsize=(13,7))
plt.title("Most Prevalent Actors/Actresses on Netflix",size='20')
sns.countplot(y = cast2, order=cast2.value_counts().index[:10], palette='Paired')
plt.show()
```

```python
actors = cast2.value_counts()
actors.head()
```

```python
usa = netflix[netflix.country == 'United States']
usa.head()
```

```python
cast_usa = usa.set_index("title").cast.str.split(',', expand=True).stack().reset_index(level=1, drop=True)
print(cast_usa)
```

```python
plt.figure(figsize=(13,7))
plt.title("Most Prevalent Actors/Actresses on Netflix in USA",size='20')
sns.countplot(y = cast_usa, order=cast_usa.value_counts().index[:10], palette='Paired')
plt.show()
```
