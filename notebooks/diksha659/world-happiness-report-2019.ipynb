```python
import warnings
warnings.filterwarnings("ignore")
import plotly.offline as py
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go
init_notebook_mode(connected = True)
from bubbly.bubbly import bubbleplot
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
report_2019 = pd.read_csv("../input/world-happiness/2019.csv")
```

```python
report_2019.head(10)
```

```python
report_2019.columns
```

```python
#correlation
corr = report_2019.corr()
plt.figure(figsize=(10,8))
sns.heatmap(corr,annot=True,cmap='Greys')
#here happiness score is highly correlated with GDP per capita, Social Support,Healthy life expectancy,quite a bit with Freedom to make life choices
```

### Freedom vs Score

```python
#Now, let's check how happiness is related with freedom and which country is happiest.
#Using bubbleplot as it is used for plotting interactive and animated bubble charts using Plotly
bubble1 = bubbleplot(report_2019,x_column='Score',y_column='Freedom to make life choices',bubble_column='Country or region',
                    color_column='Country or region',
                    size_column='GDP per capita',x_title='Happiness',y_title='Freedom',title='Happiness vs Freedom')
py.iplot(bubble1,config={'scrollzoom':True})

#Here we can see that with Freedom of making life choices, People of Finland are the most happiest and South Sudan ,
#Afghanistan comes under unhappy country.
```

### Generosity vs Score

```python
bubble2 = bubbleplot(report_2019,x_column='Score',y_column='Generosity',size_column='GDP per capita',
                    color_column='Country or region',bubble_column='Country or region',x_title='Happiness',y_title='Generosity',title=
                     'Happiness vs Generosity')
py.iplot(bubble2,config={'scrollzoom':True})

#Here people of Finland seems the happiest but less generous whereas people of Myanmar are more generous but less happy
```

### Social support vs Score

```python
bubble3 = bubbleplot(report_2019,x_column='Score',y_column='Social support',size_column='GDP per capita',
                    color_column='Country or region',bubble_column='Country or region',x_title='Happiness',y_title='Social support',title=
                     'Happiness vs Social support')
py.iplot(bubble3,config={'scrollzoom':True})
#Here People of Finland has more social support and more happy whereas people of South Sudan and Syria has less Social support and most unhappy
```

### Healthy life expectancy vs Score

```python
bubble4 = bubbleplot(report_2019,x_column='Score',y_column='Healthy life expectancy',size_column='GDP per capita',
                    color_column='Country or region',bubble_column='Country or region',x_title='Happiness',y_title='Healthy Life',title=
                     'Happiness vs Healthy Life')
py.iplot(bubble4,config={'scrollzoom':True})

#people of Finland are healthy and happy whereas people of South Sudan and Central African Republic are unhealthy and unhappy
```

#### Top 15 most generous Countries

```python
report_2019[['Country or region','Generosity']].sort_values(by = 'Generosity',ascending=False).head(20)
```

```python
#A Choropleth Map is a map composed of colored polygons. It is used to represent spatial variations of a quantity.
```

```python
#Graph represent how much generous the countries is
import plotly.graph_objects as go
import pandas as pd


fig = go.Figure(data=go.Choropleth(
    locations = report_2019['Country or region'],
    z = report_2019['Generosity'],
    text = report_2019['Country or region'],
    colorscale = 'Earth',
    autocolorscale=True,
    reversescale=True,
    marker_line_color='darkgray',
    marker_line_width=0.5,
    locationmode="country names"
))

fig.update_layout(
    title_text='Generosity',
    geo=dict(
        showframe=True,
        showcoastlines=True,
        showlakes = True,
        showocean=True,
        projection_type='hammer'
    ),
    annotations = [dict(
        x=0.55,
        y=0.1,
        xref='paper',
        yref='paper',
        text='Projection',
        showarrow = False
    )]
)

fig.show()

#We can see that China , Argentina etc are least Generous Countries
```
