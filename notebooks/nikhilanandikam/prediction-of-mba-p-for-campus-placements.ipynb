**1.Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

**2.Importing the dataset**

```python
df=pd.read_csv("/kaggle/input/factors-affecting-campus-placement/Placement_Data_Full_Class.csv")
df.head()
```

**3.Defining dependent and independent variables**

```python
x=df.iloc[:,[2,4]].values
y=df.iloc[:,12].values.reshape(-1,1)
```

**4.Splitting the dataset into the Training set and Test set in 80,20 percent**

```python
trainx,testx,trainy,testy=train_test_split(x,y,test_size=0.2)
```

**5.fitting multiple linear regression to the training set.**

```python
model=LinearRegression()
model.fit(trainx,trainy)
```

**6.predicting the test result**

```python
y_pred=model.predict(testx)
y_pred
```

```python
Final=pd.DataFrame({"Actual":testy.flatten(),"Predicted":y_pred.flatten()})
Final
```

Hence ,multi regression is completed.
