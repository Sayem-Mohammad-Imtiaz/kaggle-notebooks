```python
train_reg= pd.read_csv('/kaggle/input/heart-disease-uci/heart.csv')
train_reg
```

```python
train_reg.info()
```

```python
x = train_reg.drop('target', axis =1)
y = train_reg['target']
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.27, random_state= 100)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
scale_x_train = scale.fit_transform(x_train)
scale_x_test = scale.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
l =  LinearRegression(normalize = True)
l.fit(scale_x_train, y_train)

score_train = l.score(scale_x_train, y_train)
score_test = l.score(scale_x_test, y_test)

print('training_score:' + str(score_train))
print('test_score:' + str(score_test))
```

```python
y_pred =l.predict(scale_x_test)
y_pred
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
df
```

```python
from sklearn.svm import SVR

r = SVR(kernel='rbf')
r.fit(x_train, y_train)
score_train = r.score(x_train, y_train)
score_test = r.score(x_test, y_test)
score_train
```
