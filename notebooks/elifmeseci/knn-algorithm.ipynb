# Determination of Class

```python
sns.countplot(x="class", data=data)
data.loc[:,'class'].value_counts()
```

```python
A = data[data['class'] =='Abnormal']
N = data[data['class'] == "Normal"]
```

```python
#scatter plot
plt.scatter(A.pelvic_radius,A.sacral_slope,color="red",label="abnormal")
plt.scatter(N.pelvic_radius,N.sacral_slope,color="green",label="normal")
plt.xlabel("pelvic_radius")
plt.ylabel("sacral_slope")
plt.legend()
plt.show()
```

```python
plt.scatter(A.pelvic_radius,A.lumbar_lordosis_angle,color="red",label="abnormal")
plt.scatter(N.pelvic_radius,N.lumbar_lordosis_angle,color="green",label="normal")
plt.xlabel("pelvic_radius")
plt.ylabel("lumbar_lordosis_angle")
plt.legend()
plt.show()
```

```python
data['class'] = [1 if each == "Abnormal" else 0 for each in data['class']]
y = data['class'].values
x_data = data.drop(["class"],axis=1)
```

<a id="3"></a> <br>
# Normalization
* Normalization is one of the most important thing in KNN algorihm.
* KNN algorithm uses euclidean distance to calculate distance between points. And if we don't normalize the feature with great values, it dominates the feature with small values. 

```python
#normalization
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data))
```

<a id="4"></a> <br>
# Train Test Split

```python
#train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state=1)
```

<a id="5"></a> <br>
# KNN Model

```python
#knn model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3) # n_neighbors = k
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
```

```python
prediction
```

```python
print(" {} knn score: {}".format(3,knn.score(x_test,y_test)))
```

<a id="6"></a> <br>
# Selecting K Value

```python
#find k value
score_list = []
for each in range(1,15):
    knn2= KNeighborsClassifier(n_neighbors = each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))

plt.plot(range(1,15),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```

* as we can see if the k  value is 13. the result gives best value

```python
#knn model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 13) # n_neighbors = k
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
```

```python
prediction
```

```python
print(" {} knn score: {}".format(13,knn.score(x_test,y_test)))
```
