```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../input/cs-2021/training_set.csv")
#data
data.describe()
```

```python
data.columns
```

```python
for i in data:
    plt.figure(figsize=(15,7))
    plot_data=data.sample(1000)
    plt.scatter(plot_data['TARGET'], plot_data[i])
    plt.title(str(i))
    plt.show()
```

```python
data = data.drop(['ID', 'MAC_CODE', 'Nacelle_angle_c', 'Outdoor_temperature', 'Pitch_angle_std',], axis=1)
data.head()
```

```python
data.isnull().any()
#data[['Absolute_wind_direction_c']]
```

```python
#miss_data = []
#for var in data:
#   if data[var].isnull().any()==True:
#        miss_data.append(var)
#        data[var+'_NA']=data[var].isnull()*1 
#miss_data

data = data.fillna(method='bfill')
```

```python
#data[miss_data] = data[miss_data].fillna(data[miss_data].mean(),inplace=True)
```

```python
X = data[['Pitch_angle' ,'Absolute_wind_direction', 'Rotor_speed', 'Generator_stator_temperature',
          'Nacelle_temperature', 'Generator_speed', 'Gearbox_bearing_1_temperature_max', 
          'Hub_temperature', 'Grid_frequency', 'Pitch_angle_max']]
y = data['TARGET']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
y_pred = model.predict(X_test)
score = mean_absolute_error(y_true=y_test, y_pred=y_pred)
print('MAE score is :',score)
```
