```python
#To split the data using train_test_split with stratified parameter
from sklearn.model_selection import train_test_split
strat_train_set, strat_test_set = train_test_split(income_df, test_size=0.2, random_state=42, 
                                         stratify = income_df["income"])
```

```python
# revert to a clean training set 
# separate the predictors and the labels
train = strat_train_set.drop("income", axis=1) # drop labels for training set
train_labels = strat_train_set["income"].copy()
```

```python
# revert to a clean test set 
# separate the predictors and the labels
test = strat_test_set.drop("income", axis=1) # drop labels for test set
test_labels = strat_test_set["income"].copy()
```

#### Question 4 (Compare Models) [15 points]: 
Classify "income" by building and evaluating the following 4 types of models: Logistic Regression, Support Vector Machines, Naïve Bayes and K-Nearest Neighbors. Apply cross-validation with Pipeline, ColumnTransformer and OneHotEncoder in the modeling process. Compare the performance of different models on the Test data. Discuss your results and findings.

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler,RobustScaler,Normalizer,MaxAbsScaler


num_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="median")),
        ('std_scaler', RobustScaler()),
    ])

# df_num_tr = num_pipeline.fit_transform(df_train[['age','fnlwgt','education-num','capital-gain','capital-loss','hours-per-week']])
```

```python
cat_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy="most_frequent")),
        ("cat", OneHotEncoder())
    ])
```

```python
categorical_features
```

```python
# categorical_features.drop(" income", axis=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

num_attribs = list(continuous_features.drop("fnlwgt",axis=1))
cat_attribs = list(categorical_features.drop(["income","education"], axis=1))

#using ColumnTransformer to merge the numerical and categorical attributes
full_pipeline = ColumnTransformer([
        ("num", num_pipeline, num_attribs),
        ("cat", cat_pipeline, cat_attribs),
    ])

train_prepared = full_pipeline.fit_transform(train)
```

```python
#to get the shape of train data
train_prepared.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_validate

# log_reg=LogisticRegression()
# log_reg.fit(train_prepared,train_labels)
scores = cross_validate(LogisticRegression(),
                        train_prepared, train_labels, cv=10, scoring=('roc_auc', 'average_precision','recall'))

print("ROC_AUC: ",scores['test_roc_auc'].mean())
print("Average Precision: ",scores['test_average_precision'].mean())
print("Average Recal: ",scores['test_recall'].mean())
```

```python
# SVM
from sklearn.svm import LinearSVC, SVC
scores = cross_validate(LinearSVC(),
                        train_prepared, train_labels, cv=10, scoring=('roc_auc', 'average_precision','recall'))

print("ROC_AUC: ",scores['test_roc_auc'].mean())
print("Average Precision: ",scores['test_average_precision'].mean())
print("Average Recall: ",scores['test_recall'].mean())
```

```python
# Naive Bayes
from sklearn.naive_bayes import BernoulliNB
scores = cross_validate(BernoulliNB(),
                        train_prepared, train_labels, cv=10, scoring=('roc_auc', 'average_precision','recall'))


print("ROC_AUC: ",scores['test_roc_auc'].mean())
print("Average Precision: ",scores['test_average_precision'].mean())
print("Average Recal: ",scores['test_recall'].mean())
```

```python
# K nearest neighbors
from sklearn.neighbors import KNeighborsClassifier
scores = cross_validate(KNeighborsClassifier(),
                        train_prepared, train_labels, cv=10, scoring=('roc_auc', 'average_precision','recall'))

print("ROC_AUC: ",scores['test_roc_auc'].mean())
print("Average Precision: ",scores['test_average_precision'].mean())
print("Average Recal: ",scores['test_recall'].mean())
```
