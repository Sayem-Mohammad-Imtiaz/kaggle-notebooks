**UNIVARIATE ANALYSIS**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for col in df.iloc[:,1:3]:
    fig, ax= plt.subplots(1,3, figsize=(25,5))
    sns.boxplot(df[col], ax=ax[0])
    sns.violinplot(df[col], ax=ax[1])
    sns.distplot(df[col], ax=ax[2], label='Skewness %.2f%%'%(df[col].skew()))
    plt.legend(loc='upper left')
    plt.show()
```

1. The distributions for Height and Weight do not have any skewness or outliers
2. Maximum people have a height of 156 cm to 184cm, on a range where people have height between 140cm to 199cm with a median height of 156 cm and mean height of 169 cm
3. Maximum people have a weight of 80kg to 136 kg on a range where people have a weight between 50kg to 160kg with a median and mean weight of 106kg



**BIVARIATE ANALYSIS**

```python
fig, ax= plt.subplots(1,2, figsize=(15,8))
sns.barplot(df['Gender'],df['Height'], ax=ax[0])
sns.lineplot(df['Gender'],df['Height'], ax=ax[1])
plt.show()
```

The mean height of females is slightly more than that of males

```python
fig, ax= plt.subplots(1,2, figsize=(15,8))
sns.barplot(df['Gender'],df['Weight'], ax=ax[0])
sns.lineplot(df['Gender'],df['Weight'], ax=ax[1])
plt.show()
```

The mean weight of Males is slightly more than that of females

ANALYSIS WITH TARGET

```python
fig, ax= plt.subplots(1,2, figsize=(15,8))
sns.lineplot(df['Index'],df['Weight'], ax=ax[0])
sns.lineplot(df['Index'],df['Height'], ax=ax[1])
plt.show()
```

People who are Extremely Weak(Index 0) weigh less than required for thier height. As weight increases, their Index gets better. A mean weight of around 70kg is required for Normal (Index 2). People gaining more than that start to get Obesity (Index 4)


People who are very tall are generally Weak (Index 1) and people with shorter height are generally Overweight(Index 3). Normal (Index 2) are generally having a mean height of 174cm

```python
tab = pd.crosstab(index=df['Index'], columns=df['Gender'])
tab.plot(kind="bar", stacked=False, figsize=(15,5))
plt.xlabel('Index')
plt.ylabel('Count')
plt.legend()
plt.show()
```

1. Females are more Extremely Weak than males

2. Males are more Weak  than females

3. Females are more  Normal than males

4. Females are more Overweight than males

5. Females are more  Obese than males

6. Males are more Extreme Obese than females

MULTIVARIATE

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='Height', y='Weight', hue='Index', data=df)
plt.show()
```

1. People with equal increase in weight and height are Normal (Index 2)
2. People in lower right triangle of the plot have more height but less weight and are hence Weak(Index 1) or Extremely Weak(Index 0)
3. People in upper left triangle of the plot have more weight but less height and are hence Obese(Index 4) or Extremely Obese(Index 5)
