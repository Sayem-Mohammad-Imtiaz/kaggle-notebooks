**At the beginnig of project, take a first look at data. It's useful to get a preview.**

```python
dataFrame = pd.read_csv('/kaggle/input/world-happiness/2018.csv')
dataFrame.columns = dataFrame.columns.str.replace(' ', '_')#It's so hard to work with column names with spaces, it replaces them with "_"
dataFrame.head()
```

**And let's check the info about dataset, so learn which variables we're gonna work with.**

```python
dataFrame.info()
```

**Now let's check the correlations heat map, it's so easy to see specifically things about values. **
* ** Healthy life expectancy is positively correlated with social support.**
* **GDP per capita is positively correlated with social support. **
* **Looks like there's no negative correlationship between any parameters.**

```python
f,ax = plt.subplots(figsize=(12, 12))
sns.heatmap(dataFrame.corr(), annot = True, linewidths=.5, fmt='.1f', ax=ax)
plt.show()
```

**Let's take a look at GEnerosity and GDP per Capita's of countrys.**
* **GDP per capita has a very large range.**

```python
temp = dataFrame[['GDP per capita']]
dataFrame.Generosity.plot(kind = 'line', color = 'b',label = 'Generosity',linewidth=1,alpha = 0.7,grid = True,linestyle = '-')
plt.plot(temp,color= 'r', label = 'GDP per capita', linewidth=1, alpha=0.7,linestyle = '-')
#plt.plot(GDP_per_capita)
plt.legend(loc='upper right')
plt.title('Line Plot')
plt.show()
```

**Now it's time to look correlation with another method. We will use scatter plot to see GDP per capita and overall hapiness score**
* **It's easy to say there's an almost exponential correlation between GDP per capita and score.**

```python
dataFrame.plot(kind='scatter', x='GDP per capita', y='Score',alpha = 0.5,color = 'red')
plt.show()
```

**And it's also important to learn density of data values. Let's take a look at social support density with histogram.**
* **Looks like there are more countrys with higher social support than low social supporters.**

```python
temp1 = dataFrame[['Social support']]
plt.figure(figsize=(12,12))
plt.hist(temp1, bins=10, linewidth=2)
plt.show()
```
