```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from fbprophet import Prophet
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../input/tesla-stock-data-from-2010-to-2020/TSLA.csv')
data.head()
```

```python
data['Date'] = pd.DatetimeIndex(data['Date'])
```

```python
plt.figure(figsize=(15, 12))
plt.plot(data['Date'], data['Volume'])
```

```python
data.isnull().sum()
```

```python
p = Prophet()
data_new = data.rename(columns={'Date': 'ds', 'Volume': 'y'})
```

```python
data_new
```

```python
p.fit(data_new)
```

```python
future =  p.make_future_dataframe(periods=365, freq='D')
```

```python
future
```

```python
forecast = p.predict(future)
```

```python
plt.figure(figsize=(15, 12))
p.plot(forecast, uncertainty=True)
```

# SARIMA:

```python
data
```

```python
plt.plot(data['Date'], data['Open'])
```

```python
from statsmodels.tsa.stattools import adfuller
adfuller(data['Open'])
```

```python
data['Open'] = data['Open'].diff(2)
data.drop([0, 1, 2], axis=0, inplace=True)
```

```python
adfuller(data['Open'])
```

```python
plt.plot(data['Date'], data['Open'])
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
```

```python
model = SARIMAX(data['Open'], order=(1, 2, 1), seasonal_order=(2, 1, 2, 2))
```

```python
result = model.fit()
```

```python
result.summary()
```

```python
data['arima'] = result.fittedvalues
```

```python
plt.plot(data['Date'], data['Open'])
plt.plot(data['Date'], data['arima'])
plt.legend(['real_data', 'predict_data'])
```

```python
forecast = result.predict(start = len(data), end = len(data) + 365)
```

```python
forecast
```

```python
plt.figure(figsize=(20, 20))
plt.plot(data['Open'])
plt.plot(forecast)
plt.legend(['real_value', 'predict_value_on 2021'])
```
