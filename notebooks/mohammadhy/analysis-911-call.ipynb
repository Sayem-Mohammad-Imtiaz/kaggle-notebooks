```python

import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../input/montcoalert/911.csv')
```

```python
df.info()
```

```python
df.head(10)
```

```python
sep_title = (lambda x: x.split(':')[0])
```

```python
df['title'] = df['title'].apply(sep_title)
```

```python
df.head()
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
df['year'] = df['timeStamp'].dt.year
```

```python
df['month'] = df['timeStamp'].dt.month
```

```python
df.head()
```

```python
df['twp'].unique()
```

```python
df['twp'].sort_values(ascending=False)
```

```python
df['twp'].fillna(axis=0, method='ffill', inplace=True)
```

```python
df['zip'].fillna(axis=0, method='ffill', inplace=True)
```

```python
df.info()
```

```python
sns.countplot(x='year',data=df, hue='title')
```

```python
sns.countplot(x='month',data=df, hue='title')
```

```python
df['hour'] = df['timeStamp'].dt.hour
```

```python
sns.countplot(x='hour',data=df, hue='title')
```

```python
time_sep = lambda x: x.split(' ')[0]
```

```python
df['time'] =  df['timeStamp'].astype('str').apply(time_sep)
```

```python
df.set_index('time', inplace=True)
```

```python
df.sort_index(inplace=True)
```

```python
df
```

```python
summary = df.loc['2016-05-15':'2017-05-15']
```

```python
visualize_summary = summary.pivot_table(index='year', columns='title', values='hour', aggfunc='count')
```

```python
visualize_summary.plot(kind='barh', figsize=(6, 6))
```
