```python

ab=dt[(dt["Deaths"]<2000)]
bc=dt[(dt["Confirmed"]<2000)]
ab.Deaths.plot(kind="line",linewidth=1)
bc.Confirmed.plot(kind="line",alpha=0.5,grid=True)
plt.xlabel("Countries")
plt.ylabel("Number")
plt.show()
```

```python
a=dt[np.logical_and(dt['Deaths']<2000, dt['Recovered']<2000 )]
a.Recovered.plot()
plt.show()
```
