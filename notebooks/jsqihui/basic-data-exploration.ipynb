```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vacc = pd.read_csv("/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv")
```

## Find some facts of the data 

```python
# how many countries in the data
print("total countries: ", len(vacc.country.unique()))
print("total entries: ", vacc.shape[0])
```

```python
vacc.shape
```

```python
# find how many values are missing for each column in percentage
pd.isnull(vacc).sum() / vacc.shape[0] * 100
```

```python
# plot country based trends based on the date
def plot_trends(countries, column):
    current = vacc[vacc["country"].isin(countries)][["country", "date", column]]
    current["date"] = pd.to_datetime(current["date"])
    # print(current)
    plt.figure(figsize=(15,10))
    plt.xticks(rotation=30)
    sns.lineplot(x="date", y=column, hue="country", data=current)
```

```python
vacc["country"].unique()
```

```python
plot_trends(["Serbia", "Russia", "Brazil", "China", "Israel"], "daily_vaccinations_per_million")
```
