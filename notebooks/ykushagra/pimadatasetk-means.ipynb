```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
df = pd.read_csv("../input/pima-indians-diabetes-database/diabetes.csv")
df.head()
```

```python
df.drop(['Outcome'],axis='columns',inplace=True)
```

```python
df.head()
```

```python
sse = []
k_range = range(1,10)

for k in k_range:
    km = KMeans(n_clusters = k)
    km.fit(df)
    sse.append(km.inertia_)
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_range,sse)
```

```python
#so here k = 2 number of clusters
```

```python
km = KMeans(n_clusters=2)
yp = km.fit_predict(df)
yp
```

```python
df['cluster'] = yp
df.head()
```

```python
df.cluster.unique()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]

print(f'df1 = {df1}\n')
print(f'df2 = {df2}\n')
```

```python
plt.scatter(df1['Pregnancies'],df1['Glucose'],color='blue')
plt.scatter(df2['Pregnancies'],df2['Glucose'],color='green')
```
