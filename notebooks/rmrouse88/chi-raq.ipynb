```python
frame2=pd.read_csv('../input/Chicago_Crimes_2012_to_2017.csv')
```

```python
frame2.head()
```

```python
fig,ax = plt.subplots(figsize=(10,10))
primary = frame2.groupby("Primary Type").size().to_frame().reset_index().rename(columns={0:"Count"}).sort_values("Count",ascending=False)
primary.head()
sns.set(style="whitegrid")
sns.set_color_codes("muted")
ax = sns.barplot(x ="Count", y = "Primary Type", data = primary,ax = ax, color='b')
ax.set_title("Count of Crime by Primary Type Classification")
```

## How is the crime changing by year?##


```python
by_year=frame2.groupby("Year").size().to_frame().reset_index().rename(columns={0:"Count"}).sort_values("Count",ascending=False)
by_year.head()
```

```python
fig,ax = plt.subplots(figsize=(10,10))
sns.set(style="whitegrid")
sns.set_color_codes("muted")
ax = sns.factorplot("Year",data=frame2, kind = "count",ax =ax,color='b')
```

 - So crime seems to have been decreasing since 2012.
 - From above, we know that theft was the most prevalent crime class in the 2012-2017 data, so let's look at how it changed over time



```python
frame2[frame2["Primary Type"]=="THEFT"].groupby("Year").size().to_frame().plot()
```

 - We can see that theft followed the general trend of overall crime seen in the bar chart above
 - There was again an uptick from 2015 to 2016
 - The drop in 2017 is obviously the lack of a full year's worth of data**

```python
sixteen = frame2[frame2["Year"]==2016]
len(sixteen)
```

```python
sixteen["date"]=pd.to_datetime(sixteen["Date"],infer_datetime_format=True)
```

```python
sixteen["hour"]=sixteen["date"].apply(lambda x: x.hour)
sixteen.head()
```

```python
sixteen.groupby("hour").size().to_frame().plot()
```

## Map of Crime Occurrences from 2012 to 2017##

```python
east, west = sixteen["Longitude"].max(),sixteen["Longitude"].min()
north, south = sixteen["Latitude"].max(),sixteen["Latitude"].min()

fig,ax = plt.subplots(figsize=(10,10))
ax.set_title("2012-2017 Chicago Crime Heatmap")
m = Basemap(projection='merc', llcrnrlat=south, urcrnrlat=north,
            llcrnrlon=west, urcrnrlon=east, lat_ts=south, resolution='h')
x, y = m(sixteen['Longitude'].values, sixteen['Latitude'].values)
m.hexbin(x, y, gridsize=500,
         bins='log', cmap=cm.YlOrRd_r);
```

## Map of Theft Occurrences from 2012 to 2017 ##

```python
east, west = sixteen["Longitude"].max(),sixteen["Longitude"].min()
north, south = sixteen["Latitude"].max(),sixteen["Latitude"].min()

fig,ax = plt.subplots(figsize=(10,10))
 
m = Basemap(projection='merc', llcrnrlat=south, urcrnrlat=north,
            llcrnrlon=west, urcrnrlon=east, lat_ts=south, resolution='h',suppress_ticks=False)
x, y = m(sixteen[sixteen["Primary Type"]=="THEFT"]["Longitude"].values, sixteen[sixteen["Primary Type"]=="THEFT"]['Latitude'].values)
m.hexbin(x, y, gridsize=250,
         bins='log', cmap=cm.YlOrRd_r);
ax.set_title("2016 Chicago Theft Heatmap")
```
