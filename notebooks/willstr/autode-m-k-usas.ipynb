Antud töölehe on koostanud Tartu Ülikooli informaatika I kursuse tudengid Martin Suits ja William Straus. Töölehe eesmärgiks on arendada andmetöötlusoskust ja loovat mõtlemist. Tööleht põhineb andmestikul automudelite ja nende omaduste kohta. Uurime kolme põhilist küsimust:<br /> 
1) kuidas iseloomustada automootorite võimsust?<br /> 
2) kuidas jaotub autode kütusekulu?<br /> 
3) kas autode keskmine hind sõltub tootjast?

```python
import numpy as np
import pandas as pd
%matplotlib inline
pd.set_option('display.max_rows', 20)


df = pd.read_csv("../input/Car_sales.csv")
df
```

Järgnev tabel näitab põhistatistikat automootorite võimsuse kohta

```python
print("Keskmine võimsus(hobujõududes):", df["Horsepower"].mean())
print("Maksimaalne võimsus(hobujõududes):", df["Horsepower"].max())
print("Minimaalne võimsus(hobujõududes):", df["Horsepower"].min())

```

 **Hajusdiagramm**
Järgnev hajusdiagramm näitab tabelis olevate autode mootorisuuruste ja võimsuste suhet

```python
df.plot.scatter("Horsepower", "Engine_size", alpha=0.7);
```

**Histogramm**

Järgnev histogramm näitab autode jaotumist kütusekulu põhjal, kus ühikuks on "miles per gallon", lühendiks "mpg". Meile harjumuspäraselt on kütusekulu seda parem, mida väiksem on see arv: 30 mpg on säästlikum kui 15 mpg. 10l /100km on sama mis 23,52 mpg.

```python
df.Fuel_efficiency.plot.hist(bins=19, grid=False, rwidth=0.90); 
```

**Kokkuvõttetabel gruppide kaupa**
Järgnev tabel näitab iga automargi autokeskmist hinda USA dollarites

```python
df.groupby("Manufacturer")["Price_in_thousands"].mean().sort_values(ascending=False)*1000
```
