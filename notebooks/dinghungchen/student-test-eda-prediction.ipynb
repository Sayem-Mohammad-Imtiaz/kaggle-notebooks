## List of useful metadata

Categlorical Variables 
 - school : Name of the school the student is enrolled in
 - school_setting : The location of the school (Urban, Suburban)
 - school_type : School type (Public, Non-Public)
 - classroom: The type of classroom
 - Teaching_method : Teaching method (standard , experimental)
 - gender : the gender of the students (male,female)
 - lunch : whether a student qualifies for free/subsidized lunch 
 
Continuous Variables
 - n_student : Number of students in the class 
 - Student_id : A unique ID for each student
 - pretest : the pretest score of the students out of 100
 

Target
 - Posttest : the posttest score of the students out of 100

```python
fig = plt.figure(figsize=(20,20))
gs = fig.add_gridspec(3,3)
gs.update(wspace = 0.3 , hspace = 0.3)
sns.set(style = "darkgrid")


ax0 = fig.add_subplot(gs[0,0])
ax1 = fig.add_subplot(gs[0,1])
ax2 = fig.add_subplot(gs[1,0])
ax3 = fig.add_subplot(gs[1,1])
ax4 = fig.add_subplot(gs[2,0])
ax5 = fig.add_subplot(gs[2,1])

#Set title and label
ax0.set_title("Distribution of Top 5 School")
ax1.set_title("Distribution of school setting")
ax2.set_title("Distribution of school type")
ax3.set_title("Distribution of Teaching Method")
ax4.set_title("Distribution of Gender")
ax5.set_title("Distribution of Lunch")



sns.histplot(ax = ax0, x = target , data = df , hue = "school", hue_order = df.school.value_counts().iloc[:5].index)
sns.histplot(ax = ax1, x = target , data = df , hue = "school_setting")
sns.histplot(ax = ax2, x = target , data = df , hue = "school_type")
sns.histplot(ax = ax3, x = target , data = df , hue = "teaching_method")
sns.histplot(ax = ax4, x = target , data = df , hue = "gender")
sns.histplot(ax = ax5, x = target , data = df , hue = "lunch")
```

The above graphs shows the correlation between each categorical features vs pretest. We can see from the Top 5 schools the schools name "UKPGS" has significant better pretest results than others, also school "GOOBU" has significant worst pretest results than others. This can make a assumption where differnt schools can result student to have different test results. From the school setting, we see "suburban" are covering more on the higher tests results and urban are covering more on the lower test results. We can also see similar results happened in school type and teaching method, but in the previous graphs we determine both features are biased. Therefore we can not assume the results and relationships. Interestingly we see different results coming from "Lunch". we can see students without lunch previlage tends to have better test results than students who do have lunch previlage. 


Assumptions:

```python
# Correlation Matrix / Heat Map 

corr = df.corr()
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```
