```python
#Saving filepath into variable
ramen_ratings_filepath = "../input/ramen-ratings/ramen-ratings.csv"

#reading and storing the data in a dataframe
ramen_ratings_data = pd.read_csv(ramen_ratings_filepath, index_col=0)

#.describe() to view summary of the data
ramen_ratings_data.describe()
```
