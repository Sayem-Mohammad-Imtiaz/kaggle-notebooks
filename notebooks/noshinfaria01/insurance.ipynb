```python
import pandas as pd
```

```python
df = pd.read_csv('../input/insurance/insurance.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
model = LabelEncoder()
```

```python
df=df.apply(model.fit_transform)
```

```python
df
```

```python
x=df.iloc[:,:-1]
```

```python
x
```

```python
y=df.iloc[:,6]
```

```python
y
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.feature_selection import chi2
```

```python
fit_feature = SelectKBest(score_func = chi2,k='all')
```

```python
fit_feature.fit(x,y)
```

```python
score_col = pd.DataFrame(fit_feature.scores_,columns=['score'])
```

```python
namecol = pd.DataFrame(x.columns)
```

```python
top_feature = pd.concat([namecol,score_col],axis = 1)
```

```python
top_feature.nlargest(6,'score')
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size = .3,random_state = 1)
```

```python
xtest
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model2 = RandomForestRegressor()
```

```python
model2.fit(xtrain,ytrain)
```

```python
model2.score(xtest,ytest)
```

```python
predict = model2.predict(xtest)
```

```python
predict
```

```python
from sklearn.metrics import r2_score
score=r2_score(ytest,predict)
```

```python
score
```
