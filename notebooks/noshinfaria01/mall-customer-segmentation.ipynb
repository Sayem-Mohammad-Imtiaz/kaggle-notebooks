```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
```

```python
df = pd.read_csv('../input/customer-segmentation-tutorial-in-python/Mall_Customers.csv')
```

```python
df
```

```python
df.rename(columns={'Gender':'gender','Age':'age','Annual Income (k$)':'income','Spending Score (1-100)':'score'},inplace = True)
```

```python
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencode = LabelEncoder()
```

```python
df=df.apply(labelencode.fit_transform)
```

```python
df
```

```python
df.describe()
```

```python
seaborn.pairplot(df[['age','income','score','gender']])
```

```python
from sklearn.cluster import KMeans
```

```python
K_range = range(1,12)
wcss= []
```

```python
for k in K_range:
    km = KMeans(n_clusters=k)
    km.fit(df[['income','score']])
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.xlabel('number of cluster(k)')
plt.ylabel('sum of squared error')
plt.plot(K_range,wcss)
```

```python
import sklearn.cluster as cluster
```

```python
kmeans = cluster.KMeans(n_clusters=5)
```

```python
kmean = kmeans.fit(df[['score','income']])
```

```python
kmeans.cluster_centers_
```

```python
df['income_cluster']=kmeans.labels_
```

```python
df
```

```python
df['income_cluster'].value_counts()
```

```python
seaborn.scatterplot(x='income',y='score',hue='income_cluster',data=df)
```
