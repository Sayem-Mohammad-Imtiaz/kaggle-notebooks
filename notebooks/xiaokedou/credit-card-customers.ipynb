```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#使用pd.read_csv读取文件内容，将数据文件放入train_data中
train_data = pd.read_csv('../input/credit-card-customers/BankChurners.csv')

#使用train_data.copy,复制给train
train = train_data.copy()
```

```python
train.drop(['CLIENTNUM'], axis = 1, inplace = True)
```

```python
train['Attrition_Flag'].isnull().sum()
```

使用one-hot方式替换Attrition_Flag，当Existing Customer为1，否则为0

```python
Existing = pd.get_dummies(train['Attrition_Flag'])
```

```python
train.drop(['Attrition_Flag'], axis = 1, inplace = True)
train = pd.concat([train, Existing], axis = 1)
train['Attrition_Flag'] = train['Existing Customer']
```

删除Attrited Customer和Existing Customer

```python
train.drop(['Existing Customer'], axis = 1, inplace = True)
train.drop(['Attrited Customer'], axis = 1, inplace = True)
```

使用逻辑回归对结果进行预测

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
pred = train['Attrition_Flag']
trainTmp = train.drop(['Attrition_Flag'], axis = 1, inplace = False)
train2 = scaler.fit_transform(trainTmp)
```

```python

from sklearn.linear_model import LogisticRegression

from sklearn.neighbors import KNeighborsClassifier


from sklearn.model_selection import train_test_split

LR = LogisticRegression(C=1.0, penalty='l2', tol=0.01)
X_train, X_test, y_train, y_test = train_test_split(trainTmp, pred, test_size=0.33, random_state=42)
LR.fit(X_train,y_train)
LR.predict(X_test)
LR.score(X_test,y_test)
```
