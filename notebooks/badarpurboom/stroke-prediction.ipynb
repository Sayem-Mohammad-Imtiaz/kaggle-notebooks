# Step 2:- Preprocessing

```python
df=df.dropna()
```

```python
lab=LabelEncoder()


df['gender']=lab.fit_transform(df['gender'])
df['ever_married']=lab.fit_transform(df['ever_married'])
df['work_type']=lab.fit_transform(df['work_type'])
df['Residence_type']=lab.fit_transform(df['Residence_type'])
df['smoking_status']=lab.fit_transform(df['smoking_status'])
```

```python
X=df.drop(columns=['stroke'],axis=1)
y=df['stroke']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
std=StandardScaler()

X_train_s=std.fit_transform(X_train)
X_test_s=std.transform(X_test)
```

# Step 3 :- Model testing and accuracy


```python
# LogisticRegression

lgr=LogisticRegression()
lgr.fit(X_train_s,y_train)
out_predict=lgr.predict(X_test_s)
```

```python
print(classification_report(y_test,out_predict))
```

```python
# RandomForestClassifier

rfc=RandomForestClassifier()
rfc.fit(X_train,y_train)
out_rfc=rfc.predict(X_test)
```

```python
print(classification_report(y_test,out_rfc))
```
