```python
def random_colours(number_of_colors):
    '''
    Simple function for random colours generation.
    Input:
        number_of_colors - integer value indicating the number of colours which are going to be generated.
    Output:
        Color in the following format: ['#E86DA4'] .
    '''
    colors = []
    for i in range(number_of_colors):
        colors.append("#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)]))
    return colors
```

# Statistical Analysis

```python
data = pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/2019_nCoV_data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.describe(include="O")
```

```python
data['Last Update'] = pd.to_datetime(data['Last Update'])
```

```python
data['Day'] = data['Last Update'].apply(lambda x:x.day)
data['Hour'] = data['Last Update'].apply(lambda x:x.hour)
```

```python
data[data['Day'] == 30]
```

```python
data[data['Day'] == 30].sum()
```

* **We Can See the number of confirmed cases for Hubei Province China is 5806 on 30th in accordance with the value provided on the website which means the Confirmed Variable already included the number of people affected on previous dates that is to say no of affected cases on 28th Jan were inclusive of all cases before**
* **The value of number of deaths ,Recovered and affected are in accordance with the website when we take the data of 30th Jan that confitms that the Confirmed Variable already included the number of people affected on previous dates that is to say no of affected cases on 28th Jan were inclusive of all cases before**

```python
latest_data = data[data['Day'] == 30]
```

```python
latest_data.head()
```

```python
print('Globally Confirmed Cases: ',latest_data['Confirmed'].sum())
print('Global Deaths: ',latest_data['Deaths'].sum())
print('Globally Recovered Cases: ',latest_data['Recovered'].sum())
```

**Hence the issue is now resolved as the data is now in accordance with the Website's Information**

```python
plt.figure(figsize=(16,6))
data.groupby('Day').sum()['Confirmed'].plot()
```

**We Can See how the CoronaVirus is spreading with the days passing by**
* There is an exponential type of rise in the victims of coronovirus with time

```python
plt.figure(figsize=(16,6))
sns.barplot(x='Day',y='Confirmed',data=data)
```
