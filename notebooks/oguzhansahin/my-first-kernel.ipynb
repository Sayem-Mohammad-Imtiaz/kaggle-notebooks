```python
#Line Plot

dataFrame['Happiness.Score'].plot(kind='line',color="#a5d27b",label="Family",linewidth=2,alpha = 1,grid = True,linestyle="--") #Creating Family Plot
dataFrame['Whisker.low'].plot(kind="line",color="#64caea",label="Generosity",linewidth=2,alpha = 1,grid = True,linestyle=":")#Creating Generosity Plot
plt.xlabel('x axis')
plt.ylabel("y axis")
plt.legend(loc='upper right')
plt.title("Line Plot")
plt.show()
```

```python
#Scatter Plot

#x=Freedom y=Family

dataFrame.plot(kind="scatter",x='Freedom' ,y = 'Family', color="#89159b",alpha = 0.5)
plt.xlabel('Freedom')
plt.ylabel('Family')
plt.title('Freedom / Family Scatter Plot')
plt.legend()
plt.show()
```

```python
#Series 

series = dataFrame['Family']
print(type(series))

dataFrame1 = dataFrame[['Family']]
print(type(dataFrame1))
```

```python
# Filtering

filtered_data = dataFrame['Freedom'] > 0.6
dataFrame[filtered_data]


```

```python
#Filtering by using logical_and

dataFrame[np.logical_and(dataFrame['Family'] > 1,dataFrame['Generosity'] > 0.6)]
```

```python
# This is also same with previous code line. Therefore we can also use '&' for filtering.

dataFrame[(dataFrame['Family'] > 1)& (dataFrame['Generosity'] > 0.6)]
```

```python
# Using dictionaries in for loops :

dictionary={'Switzerland':'Western Europe','Canada':'North America','Singapore':'Southeastern Asia'}

for key,values in dictionary.items():
    print(key,' : ', values)
print('')


# For pandas we can achieve index and value with iterrows()

for index,value in dataFrame[['Family']][:2].iterrows():
    print(index , ':' , value)
                            
```
