# Comparing data from Gas and Electric Vehicles Datasets

```python
print(data['Range_Km'].mean())
print(data2['mpg'].mean())
```

It's clear from the above that gas vehicles have a higher range than electric vehicles on average. This could be attributed to the fact that gas vehicles normally have large fuel tanks, and thereby higher range.

```python
print(data['AccelSec'].mean())
print(data2[' time-to-60'].mean())
```

It's clear from the above that gas vehicles have a lower 0-60 time (on average) than electric vehicles. This could be attributed to the fact that high performance electric cars are a scarcity, and majority of the consumer vehicles choose to trade off acceleration and top speed for range.
