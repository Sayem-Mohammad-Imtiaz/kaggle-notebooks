# Lets Start Model Building

```python
model_df=df[['satisfaction_level','average_montly_hours','Work_accident','promotion_last_5years','salary']]
model_df.head()
```

> **Lets do encoding for the salary categories**

```python
dummies=pd.get_dummies(df['salary'])
dummies
```

```python
model_df=pd.concat([model_df,dummies],axis=1)
model_df.head()
```

**Now we will drop the 'medium' column for removing the encoding-trap and 'salary' columns as it's of no use now.**

```python
model_df.drop(['salary','medium'],axis=1,inplace=True)
model_df
```

```python
X=model_df
Y=df.left
```

> **Lets split the Dataset for training and Testing.**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.9)
```

> **Trainnig Model**

```python
from sklearn.linear_model import LogisticRegression
req=LogisticRegression()
req.fit(x_train,y_train)
```

> **Predincting Test Data**

```python
req.predict(x_test)
```

# Model Accuracy

```python
req.score(x_test,y_test)
```
