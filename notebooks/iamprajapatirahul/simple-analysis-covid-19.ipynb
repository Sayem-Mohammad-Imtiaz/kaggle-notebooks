## Top 10 countries with highest vaccination count

```python
cols = ['country', 'total_vaccinations']
df10 =df[cols].groupby('country').sum().sort_values('total_vaccinations', ascending=False)
df10 = df10.head(10)
df10 = df10.sort_values(by ='total_vaccinations' , ascending=True)
df10 = df10.reset_index()
```

```python
plt.plot(df10.total_vaccinations,df10.country, "*--g")
plt.ylabel("country",fontsize="15")
plt.xlabel("total_vaccinations",fontsize="15")
plt.title('countries with highest vaccination count',fontsize="20")
plt.show()
```

```python
plt.barh(df10.country, df10.total_vaccinations)
plt.xlabel("country",fontsize="15")
plt.ylabel("total_vaccinations",fontsize="15")
plt.title('countries with highest vaccination count',fontsize="20")
plt.show()
```

## Top 10 countries that are vaccinated per hundered 

```python
cols = ['country', 'total_vaccinations_per_hundred']
df10 =df[cols].groupby('country').sum().sort_values('total_vaccinations_per_hundred', ascending=False)
df10 = df10.head(10)
df10 = df10.sort_values(by ='total_vaccinations_per_hundred' , ascending=True)
df10 = df10.reset_index()
```

```python
plt.plot(df10.total_vaccinations_per_hundred,df10.country, "*--g")
plt.ylabel("country",fontsize="15")
plt.xlabel("total_vaccinations",fontsize="15")
plt.title('countries with highest vaccination count',fontsize="20")
plt.show()
```

```python
plt.barh(df10.country, df10.total_vaccinations_per_hundred)
plt.xlabel("country",fontsize="15")
plt.ylabel("total_vaccinations",fontsize="15")
plt.title('countries with highest vaccination count',fontsize="20")
plt.show()
```
