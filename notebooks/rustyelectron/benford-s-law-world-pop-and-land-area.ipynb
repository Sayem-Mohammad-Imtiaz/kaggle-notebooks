```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../input/population-by-country-2020/population_by_country_2020.csv")
```

## calculate for world population

```python
first_digits = relevant_df.iloc[:, 1].astype(str).str[:1]
```

```python
digits_list = first_digits.to_numpy()
```

```python
def relative_values(count_list):
    total_count = np.sum(count_list)
    return count_list / total_count
```

```python
def calculate_benford_statistics(digits_list):
    unique, counts = np.unique(digits_list, return_counts=True)
    data_dict = dict(zip(unique, counts))
    # calculate benford's values
    benford_values = np.log10(1 + (1/np.arange(1, 10)))
    
    # calculate relative values
    rel_vals = relative_values(counts).round(3)
    
    # error percent
    error_value = ((benford_values - rel_vals) / benford_values) * 100
    major_array = np.stack([unique, counts, rel_vals, benford_values, error_value], axis = 1)

    return pd.DataFrame(major_array, 
                        columns = ['digit','frequency','relative value', 'benford', 'error percent'])

calculate_benford_statistics(digits_list)
```

## calculate for country-wise land rea

```python
land_digits_list = relevant_df.iloc[:, 2].astype(str).str[:1].to_numpy()
land_digits_list = land_digits_list[land_digits_list != '0'] # remove zeros
```
