```python
import pandas as pd

data = pd.read_csv('../input/szeged-weather/weatherHistory.csv')
data['Formatted Date'] = data['Formatted Date'].map(lambda x: str(x)[:-6])
data['Timestamp'] = pd.to_datetime(data['Formatted Date'], format='%Y-%m-%d %H:%M:%S.000')
data['Temperature'] = data['Temperature (C)']
data.info()
data
```

```python

data['Year'] = pd.DatetimeIndex(data['Timestamp']).year
data.groupby('Year')['Temperature'].plot();
```

```python
data.groupby('Year')['Temperature'].max().plot();
data.groupby('Year')['Temperature'].mean().plot();
data.groupby('Year')['Temperature'].min().plot();
```
