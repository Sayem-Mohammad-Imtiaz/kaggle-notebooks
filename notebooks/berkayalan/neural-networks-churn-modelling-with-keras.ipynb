```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
df = pd.read_csv('../input/churn-modelling/Churn_Modelling.csv')
df.head()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df.dtypes
```

## Data Prepration

```python
X = df.iloc[:, 3:-1].values
y = df.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

Label encoding for Gender

```python
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
```

```python
X
```

We will do one hat encoding for Geography column.

```python
ohe = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ohe.fit_transform(X))
```

```python
X
```

## Splitting Training and Test Set

```python
print(X.shape)
print(y.shape)
```

```python
X_train = X[:750,:]
X_test = X[750:,:]
```

```python
y_train = y[:750]
y_test = y[750:]
```

We scale our data.

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## Creating a Neural Network with Keras

```python
nn = tf.keras.models.Sequential()
```

```python
nn.add(tf.keras.layers.Dense(units=7, activation='relu'))
nn.add(tf.keras.layers.Dense(units=7, activation='relu'))
nn.add(tf.keras.layers.Dense(units=1, activation='tanh')) #Output Layer
```

```python
nn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
nn.fit(X_train, y_train, batch_size = 64, epochs = 200)
```

## Estimating Test Results

```python
y_pred = nn.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Confusion Matrix

```python
cmatrix = confusion_matrix(y_test, y_pred)
print(cmatrix)
accuracy_score(y_test, y_pred)
```
