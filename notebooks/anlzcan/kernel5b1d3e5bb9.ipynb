```python
# alternatiflerin ve kriterlerin değerleri
x = array([[8, 7, 2,1], [5, 3, 7, 5], [7, 5, 6, 4], [9, 9, 7, 3], [11, 10, 3, 7], [6, 9, 5, 4]])

# kriterlerin weightleri(ağırlıkları)
weights = array([0.4, 0.3, 0.1, 0.2])

#Buraya kadar datalarımızı ve önemlerini atadık. Sırada 7 adımla topsis uygulanıcak.

#Adım 1: normalizasyon: cumsun() arraydeki değerlerin kümülatif toplamını üretir ve kullanılıcak ekseni belirtmek
# için ikinci bir bağımsız değişkenle de kullanılır.

col_sums = array(cumsum (x**2, 0))
norm_x = array([[round(x[i,j] / sqrt(col_sums[x.shape[0] - 1, j]),3) for j in range (4) ] for in range(6)])

#Adım 2: Her değerlendirmeyi ilgili ağırlıkla çarpmak
#wnx normalize edilmiş x matrisidir.
wnx = array([[round(weights[i] * norm_x[j,i],3) for i in range(4)] for j in range(6)])

#Adım 3: positif ve negatif ideal çözümler 
pis = array([amax(wnx[:, :1]), amax(wnx[:, 1:2]),
             amax(wnx[:, 2:3]), amax(wnx[:, 3:4])]) 
nis = array([amin(wnx[:, :1]), amin(wnx[:, 1:2]), 
                 amin(wnx[:, 2:3]), amin(wnx[:, 3:4])])

#Adım 4a: pozitif ideal solution ile aradaki mesafenin hesaplanması
b1 = array([[(wnx[j,i] - pis[i])**2 for i in range(4)] for j in range(6)])
dpis = sqrt(sum(b1,1))

#adım 4b: negatif ideal solution ile aradaki mesafelerin kıyaslanması
b2 = array([[(wnx[j,i]- nis[i])**2 for i in range (4)] for j in range(6)])
dnis = sqrt(sumb(b2,1))

#adım 5: ideal çözüme göreceli yakınlığın hesaplanması
final_solution = array ([round(dnis[i] / (dpis[i] + dnis[i]),3) for i in range (6)])
print("closeness coefficient = ", final_solution )
```
