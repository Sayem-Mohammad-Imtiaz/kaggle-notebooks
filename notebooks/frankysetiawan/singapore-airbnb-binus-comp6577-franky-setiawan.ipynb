```python
# Pada pertemuan sebelumnya, tipe data telah ditentukan beserta alasan penentuannya dan dapat dilihat pada: 
# https://binusmaya.binus.ac.id/newStudent/#/forum/reader.32D0919E-3319-4544-BD82-7EFC946B746A?id=1
# atau https://drive.google.com/file/d/1PtNsuF7CtJ3y567skaXVRXglRAPctDir/view?usp=sharing

df_lastmeet = pd.DataFrame({"Nominal": ["id", "name", "host_id", "host_name", "neighbourhood_group", "neighbourhood", "room_type"],
                            "Ordinal": ["" for i in range(7)],
                            "Interval": ["latitude", "longitude", "last_review"] + ["" for i in range(4)],
                            "Ratio": ["price", "minimum_nights", "number_of_reviews", "reviews_per_month", "calculated_host_listings_count", "availability_365", ""]})
df_lastmeet.style.hide_index()
```

```python
# Tipe data yang dapat dilakukan penghitungan mean hanya tipe data interval dan ratio.
# Begitupun penghitungan variance memerlukan mean, sehingga hanya dapat dilakukan pada tipe data interval dan ratio.
# Note: Untuk last_review, variance ditampilkan dalam int64

df_calc = df[['latitude', 'longitude', 'last_review', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
              'calculated_host_listings_count', 'availability_365']].mean(skipna=True).to_frame("Mean")
df_calc = df_calc.join(df[['latitude', 'longitude', 'last_review', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
                           'calculated_host_listings_count', 'availability_365']].var(skipna=True).to_frame("Variance"))
df_last_review = df['last_review'].dropna()
df_last_review = pd.DataFrame(pd.to_datetime(df_last_review).values.astype(np.int64))
df_last_review_mean = pd.to_datetime(df_last_review.mean())
df_last_review_var = df_last_review.var()
df_calc.loc["last_review"] = [df_last_review_mean[0], df_last_review_var[0]]
df_calc
```
