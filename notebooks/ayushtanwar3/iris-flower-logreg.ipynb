```python
df=pd.read_csv("/kaggle/input/iris-flower-dataset/IRIS.csv")
df.head()
```

```python
fig,ax=plt.subplots(figsize=(15,3), nrows=1, ncols=4, constrained_layout=True)
sns.scatterplot(x=df["sepal_length"], y=df["species"], ax=ax[0])
sns.scatterplot(x=df["sepal_width"], y=df["species"], ax=ax[1], color="red")
sns.scatterplot(x=df["petal_length"], y=df["species"], ax=ax[2], color="black")
sns.scatterplot(x=df["petal_width"], y=df["species"], ax=ax[3], color="green")
```

```python
X=df[["sepal_length", "sepal_width","petal_length", "petal_width"]]
```

```python
dummies=pd.get_dummies(df.species)
X=pd.concat([df, dummies],axis='columns')
X=X.drop(["species"], axis="columns")
y=df.species
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
X_test
```

```python
model=linear_model.LogisticRegression()
model.fit(X_train,y_train)
y_predicted=model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
cm = confusion_matrix(y_test, y_predicted)
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
