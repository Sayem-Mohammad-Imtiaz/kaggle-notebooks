```python
new_df = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
new_df.head()
```

```python
dummies=pd.get_dummies(df.salary)
dummies

df_final=pd.concat([new_df,dummies],axis='columns')
df_final=df_final.drop(["salary"],axis='columns')
df_final.head()

```

```python
X=df_final
y=df.left
```

```python
fig,ax= plt.subplots(figsize=(10,5), nrows=1, ncols=3)
sns.scatterplot(x=df["satisfaction_level"], y=df["left"], ax=ax[0], color='red')
sns.scatterplot(x=df["average_montly_hours"], y=df["left"], ax=ax[1])
sns.scatterplot(x=df["promotion_last_5years"], y=df["left"], ax=ax[2], color='green')
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y,train_size=0.001)
```

```python
model=linear_model.LogisticRegression()
model.fit(X_train,y_train)
X_test
```

```python
model.predict(X_test)
```

```python
model.predict_proba(X_test)
```

# **Accuracy of our model**

```python
print (model.score(X_test,y_test)*100,"%")
```
