```python
import numpy as np
import pylab as pl
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import wordcloud
from wordcloud import WordCloud,STOPWORDS
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from sklearn.utils import shuffle
```

```python
data = pd.read_csv('../input/impact-of-covid19-pandemic-on-the-global-economy/transformed_data.csv')
data
```

```python
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objs as go

dc = data['COUNTRY'].value_counts().reset_index()
dc.columns = [
    'COUNTRY', 
    'count'
]
dc['COUNTRY'] = dc['COUNTRY'].astype(str) + '-'
dc = dc.sort_values(['count']).tail(210)

fig = px.bar(
    dc, 
    x='count', 
    y='COUNTRY', 
    orientation='h', 
    title='COUNTRY', 
    width=1000,
    height=500 
)

fig.show()
```

```python
def impact(x):
    y = data[['CODE','COUNTRY','DATE','HDI','TC','TD','STI','POP','GDPCAP']][data["COUNTRY"] == x]
    y = y.sort_values(by="CODE",ascending=False)
    return y.head(15)
```

```python
impact("India")
```

```python
HDI= data[data['COUNTRY']=='India'].groupby(['DATE']).agg({'HDI':['sum']})
GDPCAP = data[data['COUNTRY']=='India'].groupby(['DATE']).agg({'GDPCAP':['sum']})
total1= HDI.join(GDPCAP)

TD= data[data['COUNTRY']=='India'].groupby(['DATE']).agg({'TD':['sum']})
TC = data[data['COUNTRY']=='India'].groupby(['DATE']).agg({'TC':['sum']})
total2= TD.join(TC)

POP= data[data['COUNTRY']=='India'].groupby(['DATE']).agg({'POP':['sum']})
STI = data[data['COUNTRY']=='India'].groupby(['DATE']).agg({'STI':['sum']})
total3= POP.join(STI)

plt.figure(figsize=(15,10))
plt.subplot(2, 2, 1)
total1.plot(ax=plt.gca(), title='India')
plt.ylabel("Counts", size=13)

plt.subplot(2, 2, 2)
total2.plot(ax=plt.gca(), title='India')
plt.ylabel("Counts", size=13)


plt.subplot(2, 2, 3)
total3.plot(ax=plt.gca(), title='India')
plt.ylabel("Counts", size=13)
```
