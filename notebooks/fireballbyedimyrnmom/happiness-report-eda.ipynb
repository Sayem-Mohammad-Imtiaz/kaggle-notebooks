```python
US17=df2017b.loc[df2017b['Country'] == 'United States']
#switch columns position
US17=US17[['Country','Rank','Score','GDP','Family','Life expectancy','Freedom','Corruption','Generosity']]
US17
```

```python
US18=df2018.loc[df2018['Country'] == 'United States']
#switch columns position
US18=US18[['Country','Rank','Score','GDP','Family','Life expectancy','Freedom','Corruption','Generosity']]
US18
```

```python
US19=df2019.loc[df2019['Country'] == 'United States']
#switch columns position
US19=US19[['Country','Rank','Score','GDP','Family','Life expectancy','Freedom','Corruption','Generosity']]
US19
```

```python
US20=df2020b.loc[df2020b['Country'] == 'United States']
#switch columns position
US20=US20[['Country','Rank','Score','GDP','Family','Life expectancy','Freedom','Corruption','Generosity']]
US20
```

```python
#join results for US through 2015-2020
j = [US15, US16, US17, US18, US19, US20]
USA = pd.concat(j, keys=["2015", "2016", "2017", "2018", "2019", "2020"])
```

```python
#reset indeces
USA.reset_index(level=1, inplace=True)
USA.reset_index(inplace=True)
#drop extra column
USA.drop('level_1', axis=1, inplace=True)
#rename index to year
USA.rename(columns = {"index":'Year'}, inplace = True) 
USA
```

```python
##make new
c=sns.catplot(x="Year", y='Rank', data=USA, palette ='icefire', height=4, aspect=2)
plt.xticks(rotation=50)
```

```python
sns.regplot(x='Rank', y='Generosity', data=USA, logistic=False, color='sienna')
```

```python
# gca = 'get current axis'
ax = plt.gca()

USA.plot(kind='line',x='Year',y='Rank',ax=ax)
USA.plot(kind='line',x='Year',y='Score', color='red', ax=ax)
USA.plot(kind='line',x='Year',y='Family', color='green', ax=ax)

plt.show()
```
