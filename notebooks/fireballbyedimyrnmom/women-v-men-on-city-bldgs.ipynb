The City Administrator's office compiling a list of all city owned buildings named after a man or woman to understand the representation of women in city property.    

**Task**: Visualization of the data, cleaned data for search of where woman are in bigger number and where men are in bigger number. 

**Dataset observation**:    

There are 10 rows of buildings named by **multiple people**. Those rows with multiple people named in buildings consist of **relatives** (family, spouses, etc.). Only 1 building has the name of two men only. These are **outliers** for the purpose of this notebook. 7 represent a pair on the building name, 6 were male and female pairs, 1 a male-only and 3 represent a family and do not differentiate the number of females or males.   

There are men represented in more than one building. George R. Moscone shows in 6 buildings, Harvey Milk in 3, J. M. Woods in 3. M. H. de Young in 2, and so do H. Welton Flynn and Curtis E. Green. For the purpose of this notebook, removed repeated individuals. All of those with more than one building are **male**.   

After removing 12 outliers, **41** buildings have the name of a man, **19** of a woman.    

Women are represented in most departments, but their represetation is small, compared to men. Women have low representation in those departments labeled as **RED** (there is no meta-data explaining what RED stands for), and on **ports**. Women have zero representation in the **SFMTA**, **airports**, or **PUC** (there is no meta-data explaining what PUC stands for). Women are presented the most in **Administration** buildings.

```python
# Python 3 environment with analytics libraries installed
# as defined by the kaggle/python Docker 

import numpy as np 
import pandas as pd 

#visualization libraries
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df=pd.read_csv('../input/women-representation-in-city-property-sanfrancisco/WomenRepresentaionInCityProperty-SanFrancisco.csv')
df.head()
```

```python
df.describe(include='all')
```

The columns Current Police Districts, Current Supervisor Districts,	Analysis Neighborhoods, Neighborhoods, SF Find Neighborhoods have a single value. That is, there is no difference regardless of gender in these columns. Comments column has limited data.

```python
df1=df[['Department/Source','Person', 'Gender', 'Name','Reference']]
```

```python
#find all possible values in the gender column
df1['Gender'].value_counts()
```

```python
#percentages by gender values
mylabels='Men','women', 'female and male', 'male and female', 'male and male'
mycolors='darkturquoise', 'firebrick', 'purple', 'darkorange', 'forestgreen'
plt.pie(df1['Gender'].value_counts(), 
        labels=mylabels,autopct='%1.1f%%',
       colors=mycolors)
```

```python
#a df of shared-name rows
shared = ['F & M','M & M','M & F']   
# selecting rows based on condition 
shared_df = df1[df1['Gender'].isin(shared)] 
#reset indexes
shared_df=shared_df.reset_index(drop=True)
shared_df
```

```python
#percentages by gender values
mylabels='female and male', 'male and female', 'male and male'
mycolors='gray', 'blueviolet', 'royalblue'
plt.pie(shared_df['Gender'].value_counts(), 
        labels=mylabels,autopct='%1.1f%%',
       colors=mycolors)
```

There are 10 rows of buildings named by **multiple people**. Those rows with multiple people named in buildings consist of **relatives** (family, spouses, etc.). Only 1 building has the name of two men only. These are **outliers** for the purpose of this notebook. 7 represent a pair on the building name, 6 were male and female pairs, 1 a male-only and 3 represent a family and do not differentiate the number of females or males.

```python
# delete rows with multiple people on a building 
df2 = df1[ (df1['Gender'] != 'M & F') & (df1['Gender'] != 'M & M') & (df1['Gender'] != 'F & M')]
```

```python
df2.Gender.value_counts()
```

```python
#find duplicates in rows based on 'Person' column 
repeat1 = df2[df2.duplicated('Person')] 
```

George R. Moscone shows in 6 buildings, Harvey Milk in 3, J. M. Woods in 3. 
M. H. de Young in 2, and so do H. Welton Flynn and Curtis E. Green. For the purpose of this notebook, I will remove repeated individuals. All of those with more than one building are **male**. Those buildings mostly relate to **transportation** as seen in the bar plot.                 

```python
df3=df2.drop([2, 22, 31, 38, 42, 46, 50, 52, 70, 76, 77, 80])
#then, reset index
df3=df3.reset_index(drop=True)
df3
```

```python
df3.describe(include='all')
```

The dataset (df3) of individuals on city buildings contains 60 entries. That is 12 less than the original dataframe (df). **41 are men, 19 are women**. Without the rows with NaNs in the Reference column, the dataset (df3) shriks to 38 rows total (28 men, 11 women). I left those in the dataframe to keep a larger population in the data for interpretation.    

```python
#df3
df3.groupby('Department/Source')['Department/Source'].agg('count').plot(kind='pie',title='Group-By Department/Source', figsize=(7,7), autopct='%0.1f')
```

```python
fig, ax = plt.subplots(figsize=(10,5))
df3.groupby(['Department/Source']).count()['Gender'].plot(ax=ax, color='purple')
```

```python
mycolors='salmon', 'turquoise'
df_gn = df3.groupby(['Department/Source', 'Gender']).size().unstack()
df_gn.plot(kind = 'bar', color=mycolors, title='Departments by gender')
```
