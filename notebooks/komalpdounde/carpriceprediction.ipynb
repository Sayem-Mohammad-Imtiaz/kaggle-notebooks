```python
x=dt.drop(columns=['price'])
y=dt[['price']]
```

```python
plt.figure(figsize=(25,16))
sns.heatmap(x.corr(),annot=True)
```

```python
#### with the dollowing function we can select highly correlated features


def correlation(dataset, threshold):
    col_corr=set()
    corr_matrix=dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>threshold:
                colname=corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features=correlation(x,0.85)
corr_features
```

```python
x1=x.drop(columns=corr_features)
x1.head()
```

```python
from sklearn.feature_selection import f_regression
```

```python
p_values=f_regression(x1.select_dtypes(exclude=object),y)[1]
p_values=p_values.round(3)
p_values
```

```python
cols=x1.select_dtypes(exclude=object).columns
cols
```

```python
p_value=pd.DataFrame({'features':cols})
p_value['p_values']=p_values
p_value
```

```python
p_value[p_value['p_values']>0.05]
```

```python
### as this feature has p value > 0.05 we need to discard all these features
```

```python
x2=x1.drop(columns=['car_ID','symboling','carheight','stroke','compressionratio','peakrpm'])
x2
```

```python
x3=pd.get_dummies(x2,drop_first=True)
x3
```

```python
p_values=f_regression(x3,y)[1]
p_values=p_values.round(3)
p_values
```

```python
p_df=pd.DataFrame({'features':x3.columns})
p_df['p_value']=p_values
p_df
```

```python
drop=p_df[p_df['p_value']>0.05]
drop
```

```python
drop_cols=[]
for i in drop['features']:
    drop_cols.append(i)
```

```python
drop_cols
```

```python
x4=x3.drop(columns=drop_cols)
```
