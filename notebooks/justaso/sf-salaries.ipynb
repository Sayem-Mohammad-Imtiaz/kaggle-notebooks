```python
sal.groupby('Year')['TotalPay'].mean() #mean value of the TotalPay by year
```

```python
sal['JobTitle'].nunique() #number of unique job titles
```

```python
sal['JobTitle'].value_counts().head() #top 5 most common job titles and how many people have said titles
```

```python
sum(sal[sal['Year']==2013]['JobTitle'].value_counts()==1) #number of job titles represented by one person only in 2013
```

```python
sum(sal['JobTitle'].apply(lambda x: 'chief' in x.lower().split())) #number of people that have the word 'chief' in their job title
```

```python
sal['title_len'] = sal['JobTitle'].apply(len) #creates new column with the length of the job title
```

```python
sal[['title_len','TotalPayBenefits']].corr()
```

There does not seem to be any correlation between the total pay including benefits and the length of the job title.
