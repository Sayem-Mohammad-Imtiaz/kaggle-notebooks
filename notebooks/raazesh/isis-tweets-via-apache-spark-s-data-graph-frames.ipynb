# Exploratory Analysis of ISIS Tweets Data 
## Via Apache Spark DataFrames and GraphFrames

### 2016, Raazesh Sainudiin and Rania Sahioun

We explore the extracted *mentions network* using: 

 -  connected components, strongly connected components
 - community detection via label propagation
 - page rank
 - and various SparkSQL queries in DataFrame

This is part of *Project MEP: Meme Evolution Programme* and supported by databricks academic partners program.

The analysis is available in the following databricks notebook:

- [http://lamastex.org/lmse/mep/fighting-hate/extremist-files/ideology/islamic-state/tweetsIsis.html](http://lamastex.org/lmse/mep/fighting-hate/extremist-files/ideology/islamic-state/tweetsIsis.html)

For details on the mathematical model motivating the exploratory data analysis in the above notebook see:
 - The Transmission Process: A Combinatorial Stochastic Process for the Evolution of Transmission Trees over Networks, Raazesh Sainudiin and David Welch, Journal of Theoretical Biology DOI: 10.1016/j.jtbi.2016.07.038 (In Press) [preprint PDF 900KB](http://lamastex.org/preprints/20160806_transmissionProcJTBFinal.pdf).
