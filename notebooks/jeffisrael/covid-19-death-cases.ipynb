```python
# Important models and tools to import

from sklearn.neighbors import KNeighborsClassifier
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split as tts
from sklearn.metrics import accuracy_score
import tensorflow as tf
print(tf.__version__)
import math
```

```python
death_cases = pd.read_csv('../input/novel-corona-virus-2019-dataset/time_series_covid_19_deaths.csv')
print(death_cases)
print(death_cases.columns)
```

```python
features = ['1/23/20','1/23/20','1/24/20','1/25/20','1/26/20', '1/27/20', '8/18/20', '8/19/20', '8/20/20', '8/21/20', '8/22/20', '8/23/20', '8/24/20']
x = death_cases[features]
target = ['8/25/20']
y = death_cases[target]

x_train, x_test, y_train, y_test = tts(x, y, test_size=0.25)

clf = KNeighborsClassifier()

clf.fit(x_train, y_train)

prediction = list(clf.predict(x_test))

z = accuracy_score(y_test, prediction)


print('The predicted death cases are '+ str(prediction))
print('Actual Data: ', str(y_test))

plt.plot(prediction)

accuracy = math.floor(z*100)  
print('Accuracy: ' + str(accuracy) + '%')

if accuracy < 50:
    print('This prediction is almost inaccurate! ')
    
if accuracy > 50 and accuracy < 70:
    print('Nice prediction. Try improving it! ')
    
if accuracy > 70:
    print('Excellent prediction! Good job!')
    
print("Here's the graph...")
```
