```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

# **Avocado Prices and volume in the state of New York: organic and conventional**

```python
my_filepath = "../input/avocado-prices/avocado.csv"
my_data = pd.read_csv(my_filepath, index_col="Unnamed: 0")
```

```python
my_data = my_data[my_data['region']=='NewYork']
my_data['Date']=pd.to_datetime(my_data.Date)
my_data=my_data.sort_values(by="Date")
my_data_org=my_data[my_data['type']=='organic'].set_index('Date')
my_data_con=my_data[my_data['type']=='conventional'].set_index('Date')
```

```python
my_data_org.head()
```

# Avocado prices

**Is the gap getting closer?**

```python
# Create a plot
plt.figure(figsize=(14,5))

sns.lineplot(data=my_data_org["AveragePrice"],label="Organic")
sns.lineplot(data=my_data_con["AveragePrice"],label="Conventional")

plt.xlabel("Date")
plt.title('Avocado prices')
```

In the plot above we can see that the organic avocado prices are higher than the conventional prices, but the gap it's getting smaller

# Avocado volume

**Which type of product is more consumed in the state of New York ???**

```python
plt.figure(figsize=(14,5))

sns.lineplot(data=my_data_org["Total Volume"],label="Organic")
sns.lineplot(data=my_data_con["Total Volume"],label="Conventional")

plt.xlabel("Date")
plt.title('Avocado volume sold')
```

Here we can see that the consumption of conventional avocado is greater than the consumption of organic avocado in the estate of New York
