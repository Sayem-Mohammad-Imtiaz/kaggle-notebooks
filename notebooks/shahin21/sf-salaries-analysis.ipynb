**How many people have the word Chief in their job title? (This is pretty tricky)**

```python
def chief_string(title):
    if 'chief' in title.lower().split():
        return True
    else:
        return False
```

```python
sum(sal['JobTitle'].apply(lambda x:chief_string(x)))
```
