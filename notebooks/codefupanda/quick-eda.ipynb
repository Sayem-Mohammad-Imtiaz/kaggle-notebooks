```python
import matplotlib.pyplot as plt

import seaborn as sns

%matplotlib inline
plt.rcParams['figure.dpi'] = 100
plt.rcParams['figure.dpi'] = 100
sns.set_style('darkgrid')
```

```python
data = pd.read_csv('/kaggle/input/factors-affecting-campus-placement/Placement_Data_Full_Class.csv')
data = data.drop('sl_no', axis=1)
```

```python
data.head()
```

```python
data.isna().any()
```

```python
# Data looks clean, we dont have null salary where status is placed
data[data['status'] == 'Placed']['salary'].isna().any()
```

```python
# ~70% of people are placed
status_ = data['status'].value_counts()
status_ = status_/ len(data['status'])
status_.plot(kind='bar')
```

```python
# Most of the salaries range from 20K to ~40K
# And of course few outliers
pd.cut(data['salary'], bins=5).value_counts()
```

## How Important is work experience 


```python
workex_placed = data.groupby(['workex', 'status'])['status'].count().unstack('status')
workex_placed
```

```python
# Having prior work experience helps in getting job
# Observe that overall placed rate is 70%, if student does not have workex chance of getting placed falls to 60%
workex_placed.div(workex_placed.sum(axis=1), axis=0).plot(kind='bar')
```

## How Important is HSC Subject

```python
# Commerce and Science seems to align well with overall placed rate, arts student lag behind
workex_placed = data.groupby(['hsc_s', 'status'])['status'].count().unstack('status')
workex_placed.div(workex_placed.sum(axis=1), axis=0).plot(kind='bar')
```

```python
# Commerce and Science seems to align well with overall placed rate, arts student lag behind
workex_placed = data.groupby(['specialisation', 'status'])['status'].count().unstack('status')
workex_placed.div(workex_placed.sum(axis=1), axis=0).plot(kind='bar')
```

```python
fig, axes = plt.subplots(2, 3, figsize=(10, 5))

df = data.groupby(['ssc_p_label', 'status'])['status'].count().unstack('status')
df.div(df.sum(axis=1), axis=0).plot(kind='bar', ax=axes[0, 0], legend=None)

df = data.groupby(['hsc_p_label', 'status'])['status'].count().unstack('status')
df.div(df.sum(axis=1), axis=0).plot(kind='bar', ax=axes[0, 1], legend=None)

df = data.groupby(['degree_p_label', 'status'])['status'].count().unstack('status')
df.div(df.sum(axis=1), axis=0).plot(kind='bar', ax=axes[1, 0], legend=None)

df = data.groupby(['mba_p_label', 'status'])['status'].count().unstack('status')
df.div(df.sum(axis=1), axis=0).plot(kind='bar', ax=axes[1, 1], legend=None)

df = data.groupby(['etest_p_label', 'status'])['status'].count().unstack('status')
df.div(df.sum(axis=1), axis=0).plot(kind='bar', ax=axes[0, 2], legend=None)

fig.tight_layout()
```
