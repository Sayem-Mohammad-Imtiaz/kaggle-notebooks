```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, RobustScaler
from sklearn.compose import ColumnTransformer
from sklearn import set_config; set_config(display='diagram')
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv('/kaggle/input/predict-test-scores-of-students/test_scores.csv')
data.head()
```

```python
data.columns
```

```python
non_num_features = ['school', 'school_setting', 'school_type',
       'teaching_method', 'n_student', 'gender', 'lunch',
       'pretest']
num_features = ['n_student', 'pretest']
```

```python
X = data.drop(['classroom', 'student_id', 'posttest'], axis=1)
y = data['posttest']
```

```python
num_transformer = Pipeline([('scaler', RobustScaler())])
cat_transformer = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer([
    ('num_transformer', num_transformer, num_features),
    ('cat_transformer', cat_transformer, non_num_features)])
```

```python
preprocessor
```

```python
final_pipe = Pipeline([
    ('preprocessor', preprocessor),
    ('linear_regression', LinearRegression())
])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.2, random_state = 6)
```

```python
final_pipe.fit(X_train, y_train)
```

```python
cross_val_score(final_pipe, X_test, y_test, cv=5, scoring='r2').mean()
```

```python
y_pred = final_pipe.predict(X_test)
df_compare = pd.DataFrame(y_pred, y_test, columns=['predicted']).reset_index()
df_compare.head()
```
