## Model

You can also check the outliers here, it is a good practice. Now we will make our model. 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

*LinearRegression* is for our modeland *train_test_split* is for splitting the main data for training and testing operations.[](http://)

We will create a X variable which will be our main part to see the pattern.

I choose to drop the column date because it is a categorical data.

y variable is called the target variable.

```python
X = data.drop(columns = ["price","date"])
y = data[["id","price"]]
```

Now we are going to split our data.

```python
train_X, test_X, train_y,test_y = train_test_split(X,y)
```

To build a model you should do the following operations.

```python
model = LinearRegression()
```

Fitting is one of the significant part of our model. It finds the patterns inside the data and build a model acording to the coefficients of the pattern.

```python
model.fit(train_X,train_y)
```

Let's see the training score of our model.
Remember the score gives us the R^2 score.

```python
model.score(train_X,train_y)
```

Now we're going to test our test_X data.

```python
pred = model.predict(test_X)
```

Let's see our predictions.

```python
pred
```

For the final step we will see the score of our test data.

```python
model.score(test_X,test_y)
```

Awesome! You built a Linear Regression Model! Congratulations :) 

*Note: R^2 score means that you guessed the test_X data %85 correctly.*
