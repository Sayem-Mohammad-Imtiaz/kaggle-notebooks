# Heart Disease Prediction with XGBoost

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
import xgboost as xgb
```

## Preprocessing

```python
test = data[:30]
test_X = test.drop("target",1)
test_y = test["target"]
data = data[30:]
X = data.drop("target",1)
y = data["target"]
```

## Model Building

```python
model = xgb.XGBClassifier(learning_rate=0.5,n_estimators=50,max_depth=4)
model.fit(X,y)
```

```python
pred = model.predict(test_X)
print("Model accuracy: ",accuracy_score(test_y,pred)*100)
```
