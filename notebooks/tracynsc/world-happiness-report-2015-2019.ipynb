**Create a dataframe to compare 2015 and 2019 values**

```python
report1519 = report19.merge(report15, left_on = "Country", right_on = "Country", suffixes=('_19', '_15'))
```

```python
report1519["change"] = report1519["HappinessScore_19"] - report1519["HappinessScore_15"]
```

```python
report1519.sort_values(by = "change", ascending = False).head(3)
```

Let's see which Region has the most country with a decrease in Happiness Score from 2015-2019 period

```python
decrease = report1519[report1519["change"] < 0].groupby("Region_15").change.count().reset_index()
```

```python
allcountry = report1519.groupby("Region_15").change.count().reset_index()
```

```python
merged = allcountry.merge(decrease, left_on = "Region_15", right_on = "Region_15", suffixes = ("_all", "_dec"))
merged["perc"] = round(merged["change_dec"] / merged["change_all"] * 100, 1)
merged["change_inc"] = merged["change_all"] - merged["change_dec"]
```

```python
merged.plot(x = "Region_15", y = ["change_inc", "change_dec"], kind = "bar", stacked = True,
        title = "Change in Happiness Score")
```

Western Europe is the best performing region: it has the lowest percentage of countries observing a decrease in happiness score

**Using the latest data, 2019, these are the top 10 happiest countries.**

# **DATA MODELING**

```python
import statsmodels.formula.api as smf
```

```python
happydummy = pd.get_dummies(aggreport[["Region"]], drop_first = True)
```

```python
happydummy
```

```python
aggreport.Region.unique()
```

```python
model1 = smf.ols(formula = "HappinessScore ~ Region + Year + GDP + Family + Life + Freedom + Trust + Generosity", data = aggreport).fit()
```

```python
print(model1.summary())
```

Note that the p-values for Region[Latin America and Carribean], Region[North America] and Region[Western Europe] are large. We will drop the Region variable.

```python
model2 = smf.ols(formula = "HappinessScore ~ Year + GDP + Family + Life + Freedom + Trust + Generosity", data = aggreport).fit()
print(model2.summary())
```

Model 2 has an R^2 value of 77%, about 4% less than Model 1.

Notes [2] indicate there might be multicollinearity issue, let's drop the Year variable in Model 3.

```python
model3 = smf.ols(formula = "HappinessScore ~ GDP + Family + Life + Freedom + Trust + Generosity", data = aggreport).fit()
print(model3.summary())
```

By dropping the Year variable, we can see that the p-value for Generosity decreased, which is an improvement. Hence we will use Model 3 as our final model.
