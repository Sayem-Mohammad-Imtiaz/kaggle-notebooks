```python
import pandas as pd
import numpy as np
import seaborn as sn
import plotly.express as px
import plotly.graph_objects as go
from plotly.offline import iplot
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pd.set_option('display.max_columns',None)
df_comp = pd.read_csv('/kaggle/input/indian-school-education-statistics/percentage-of-schools-with-comps-2013-2016.csv')
df_comp.head()
```

### Firstly,lets visualize the 'All Schools' Column with respect to the states/ut

```python
top_list = df_comp.sort_values(['All Schools'],ascending=False)
top_list.head(2)
```

# Let's Visualize using BarPlot

```python
px.bar(data_frame=top_list,x = 'State_UT',y = 'All Schools',labels = {'x':'State and UT','y':'All Schools'},opacity=0.8,color_discrete_sequence=['purple'])
```

### from the bar graph above we can see that union territories like Lakshadweep,Puducherry and Chandigarh have contributed the most towards the "computerized" schools,however these have small populations,so if we have a look at the states,Kerala has contributed the most towards the idea of "computerised" schools,and states like Bihar and Jharkhand have not many Contributions:(

# Let's Visualise using PIE CHART

```python
chart = px.pie(data_frame=top_list,values = 'All Schools',names='State_UT',height = 600)
chart.update_traces(textposition = 'inside',textinfo = 'percent+label')

chart.update_layout(title_x = 0.5,
                   geo = dict(showframe = False,
                             showcoastlines = False))
chart.show()
```

### Now lets look section by section of schools of states...i.e Primary,Secondary,SeniorSecondary etc...and their contributions to the "computrised" schools..

### Let's first look at the Four Basic schools,Primary,Secondary,UnderPrimary and SeniorSecondary

```python
df = df_comp.copy()
df.head(2)
```

```python
x = df.State_UT

trace_1 = {
    'x':x,
    'y':df.Primary_Only,
    'name':'Primary_Education',
    'type':'bar'
};
trace_2 = {
    'x':x,
    'y':df.Sec_Only,
    'name':'Secondary_Education',
    'type':'bar'
};
trace_3 = {
    'x':x,
    'y':df.HrSec_Only,
    'name':'HigherSecondary',
    'type':'bar',
};
trace_4 = {
    'x':x,
    'y':df.U_Primary_Only,
    'name':'UnderPrimary',
    'type':'bar',
};
data = [trace_1,trace_2,trace_3,trace_4]
layout = {
    'xaxis':{'title':'Computing in Eductaion'},
    'barmode':'relative',
    'title':'Trend Of computing Education in Indian States',
}
fig = go.Figure(data = data,layout=layout)
iplot(fig)
```

### As we can see from the graph that kerala has most number of "computrised" schools in coparision to other States and in Union territories Daman and Diu has given a huge contribution...but hang on!...there're even sub branches...like UnderPrimary with Secondary and stuff...so lets check out all the contributions given by states and union terriotories towards "computrised" schools

```python
x = df.State_UT

trace_1 = {
    'x':x,
    'y':df.Primary_Only,
    'name':'Primary_Education',
    'type':'bar'
};
trace_2 = {
    'x':x,
    'y':df.Sec_Only,
    'name':'Secondary_Education',
    'type':'bar'
};
trace_3 = {
    'x':x,
    'y':df.HrSec_Only,
    'name':'HigherSecondary',
    'type':'bar',
};
trace_4 = {
    'x':x,
    'y':df.Primary_with_U_Primary,
    'name':'UnderPrimary',
    'type':'bar',
};
trace_5 = {
    'x':x,
    'y': df.Primary_with_U_Primary_Sec,
    'name':'PrimarywithSecondary',
    'type':'bar',
};
trace_6 = {
    'x':x,
    'y':df.Primary_with_U_Primary_Sec_HrSec,
    'name':'Primary with SeniorSecondary',
    'type':'bar'
};
trace_7 = {
    'x':x,
    'y':df.U_Primary_Only,
    'name':'UnderPrimary',
    'type':'bar',
};
trace_8 = {
    'x':x,
    'y':df.U_Primary_With_Sec,
    'name':'UnderPrimaryWithSecondary',
    'type':'bar'
};
trace_9 = {
    'x':x,
    'y':df.U_Primary_With_Sec_HrSec,
    'name':'UnderPrimarywithSeniorSecondary',
    'type':'bar'
};
data = [trace_1,trace_2,trace_3,trace_4,trace_5,trace_6,trace_7,trace_8,trace_9]
layout = {
    'xaxis':{'title':'Computing in Eductaion'},
    'barmode':'relative',
    'title':'Trend Of computing Education in Indian States',
}
fig = go.Figure(data = data,layout=layout)

iplot(fig)



```

### If we see with respect to all sections of schools i.e underprimary,underprimary with secondary etc etc...we find that Kerala has the most number of "Computrised" schools as compared to any other states...and states like Gujarat and Punjab have also done a preety good job ...as these states have developed a lot since past 3 years..we can see that these states have a good improvement in their education sector,by providing all the necessary contributions and Union territories like PuduCherry and Daman and Diu have also contrbuted in a good number,

### States like Mizoram,Bihar and JharKhand must develop in this Sector as there cntributions towards "Computrised" schools is about ~50%

### All said and prooved by the data itself,that Kerala still remains the State which Provides the Most number of literates and contributes immensly towards the "computrised" schools
