# math score of students analysis

```python

sns.kdeplot(x=data['math score'],color='red',hue=sorted(data['parental level of education']),fill=True)
```

# **performance of students based on their lunch type**

```python
data1[data1['lunch']=='standard'].describe()
```

```python
data1[data1['lunch']=='free/reduced'].describe()
```

```python
mss = data1[data1['lunch']=='standard']['math score']
msf = data1[data1['lunch']=='free/reduced']['math score'] 
```

```python
rss = data1[data1['lunch']=='standard']['reading score']
rsf = data1[data1['lunch']=='free/reduced']['reading score']
```

```python
plt.figure(figsize=[10,8])
plt.subplot(2,2,1)
plt.title('math score vs standard lunch')
sns.boxplot(y=mss,)
plt.subplot(2,2,2)
plt.title('math score vs fee/reduced lunch')
sns.boxplot(y=msf)
plt.subplot(2,2,3)
plt.title('reading score vs fee/reduced lunch')
sns.boxplot(y=rss)
plt.subplot(2,2,4)
plt.title('reading score vs fee/reduced lunch')
sns.boxplot(y=rsf)
```

```python
wss = data1[data1['lunch']=='standard']['writing score']
wsf = data1[data1['lunch']=='free/reduced']['writing score']
plt.figure(figsize=[10,8],)
plt.subplot(2,2,1)
plt.title('writing score vs fee/reduced lunch')
sns.boxplot(y=wsf)

plt.subplot(2,2,2)
plt.title('writing score vs standard lunch')
sns.boxplot(y=wss)
```
