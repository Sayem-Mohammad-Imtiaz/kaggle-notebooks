```python
#converting the values using Label Encoding
```

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
le = LabelEncoder() 
```

```python
df['Sex']= le.fit_transform(df['Sex'])
```

```python
df['BP'] = le.fit_transform(df['BP'])
```

```python
df['Cholesterol']=le.fit_transform(df['Cholesterol'])
```

```python
#spliting the data as training and testing set
```

```python
df.columns
```

```python
X = df[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']]
```

```python
y = df[['Drug']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state=0)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
#applying the model 
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
# checking the accuracy of the model using confusion Matrix & metrics 
```

```python
from sklearn import metrics
```

```python
from sklearn.metrics import confusion_matrix
```

```python
metrics.confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
metrics.accuracy_score(y_test,y_pred)
```

```python
# changing the model as XGBoost Classifier 
```

```python
from xgboost import XGBClassifier
```

```python
model_xg = XGBClassifier()
```

```python
model_xg.fit(X_train,y_train)
```

```python
pred = model_xg.predict(X_test)
```

```python
#accuracy Score
```

```python
metrics.accuracy_score(y_test,pred)
```
