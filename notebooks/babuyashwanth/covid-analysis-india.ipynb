**seaborn** 

```python
plt.figure(figsize=(15,10))
sns.relplot(data=case,x='Total Cases',y='Discharged')
```

```python
plt.figure(figsize=(15,10))
sns.barplot(x='Total Cases',y='State/UTs',data=case)
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.scatterplot(x='State/UTs',y='Active',data=case,s=100, color=".3", marker="*")
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.scatterplot(x='State/UTs',y='Deaths',data=case)
```
