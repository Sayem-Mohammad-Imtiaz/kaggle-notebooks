```python
# Check Differ from "AVG-Entrepreneurship Index"
ent_mean = lambda x : round(x - x.mean(),2)
df['Differ Ent-AVG'] = df["Entrepreneurship Index"].transform(ent_mean)
df.head()
```

```python
# Top 5 "Entrepreneurship Index	" in the both developed and developing countries groups in the same table

developed_df = df.loc[df["Level of development"] == "Developed"]
Rt5_df = developed_df.sort_values(by = "Entrepreneurship Index" , ascending = False)[0:5]
developing_df = df.loc[df["Level of development"] == "Developing"]
Lt5_df = developing_df.sort_values(by = "Entrepreneurship Index" , ascending = False)[0:5]
```

```python
top5_ent = pd.concat([Rt5_df , Lt5_df])
top5_ent
```

```python
# Analyzing Data based on countries
cgrp_df = df.groupby("Country")
cgrp_df.aggregate({"Inflation rate":"count" , "Women Entrepreneurship Index":"mean" }).sort_values(by 
                                                                                                   = "Women Entrepreneurship Index"
                                                                                                   , ascending = False)
```
