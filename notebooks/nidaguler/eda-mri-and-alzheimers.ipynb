```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
df_sec=pd.read_csv("../input/mri-and-alzheimers/oasis_cross-sectional.csv")
df_long=pd.read_csv("../input/mri-and-alzheimers/oasis_longitudinal.csv")
```

```python
df_sec.drop(["Educ","SES","MMSE","CDR","Delay","ID"],axis=1,inplace=True)
```

```python
df_sec.columns=["gender","hand","age","eTIV","nWBV","ASF"]
```

```python
sns.set(style='whitegrid')
ax=sns.barplot(x=df_sec['gender'].value_counts().index,y=df_sec['gender'].value_counts().values,palette="Blues_d",hue=['F','M'])
plt.legend(loc=8)
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Show of Gender Bar Plot')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x = "hand", y = "eTIV", hue = "gender", data = df_sec)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x = "hand", y = "nWBV", hue = "gender", data = df_sec)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x = "hand", y = "ASF", hue = "gender", data = df_sec)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(15,7))
ax = sns.pointplot(x="eTIV", y="nWBV", hue="gender",data=df_sec)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,7))
ax = sns.pointplot(x="eTIV", y="ASF", hue="gender",data=df_sec)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,7))
ax = sns.pointplot(x="eTIV", y="age", hue="gender",data=df_sec)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(4,4))
labels=['eTIV', 'nWBV', 'ASF']
colors=['blue','red','yellow']
explode=[0,0,0.1]
values=[df_sec.eTIV.mean(),df_sec.nWBV.mean(),df_sec.ASF.mean()]

plt.pie(values,labels=labels,colors=colors,explode=explode,autopct='%1.1f%%',shadow=True)
plt.legend(['eTIV', 'nWBV', 'ASF'] , loc=3)
plt.axis('equal')
plt.tight_layout()
plt.show()
```
