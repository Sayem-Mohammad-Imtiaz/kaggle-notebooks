```python
X = dataset.iloc[:, [1,3,4,5,6,7,8,9,10,12]].values
y = dataset.loc[:, 'selling_price']
```

```python
y = np.array(y)
```

```python
X[:, 6] = [i.replace('kmpl', "").strip() for i in X[:, 6]]
```

```python
X[:, 6] = [i.replace('km/kg', "").strip() for i in X[:, 6]]
```

```python
X[:, 7] = [i.replace('CC', "").strip() for i in X[:, 7]]
```

```python
X[:, 8] = [i.replace('bhp', "").strip() for i in X[:, 8]]
```

```python
X[:, 6] = np.array(X[:, 6], dtype='float')
X[:, 7] = np.array(X[:, 7], dtype='float')
X[:, 8] = np.array(X[:, 8], dtype='float')
X[:, 6] = np.array(X[:, 6], dtype='int')
X[:, 7] = np.array(X[:, 7], dtype='int')
X[:, 8] = np.array(X[:, 8], dtype='int')
X[:, 9] = np.array(X[:, 9], dtype='int')
```

```python
X = pd.DataFrame(X)
X
```

```python
plt_features = X.iloc[:, [0,1,6,7,8,9]]
```

```python
for i in plt_features:
    plt.scatter(plt_features[i], y)
    plt.title('Correlation')
    plt.xlabel(i)
    plt.ylabel('Selling Price')
    plt.show()
```

```python
X = pd.get_dummies(X, columns=[2,3,4,5], drop_first=True)
X
```

```python
X = np.array(X)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 300).fit(X_train, y_train)
print(regressor.score(X_train, y_train))
```

```python
y_pred = regressor.predict(X_test)
y_pred = np.array(y_pred, dtype='int')
y_test = pd.Series(y_test)
y_pred = pd.Series(y_pred)
```
