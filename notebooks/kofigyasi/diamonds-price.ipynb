```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(categories=[['Fair', 'Good', 'Very Good', 'Ideal', 'Premium'], ['J', 'I', 'H', 'G', 'F', 'E', 'D'], 
                                     ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']])
X[:, 1:4] = encoder.fit_transform(X[:, 1:4])
```

```python
X = pd.DataFrame(X)
X
```

```python
X = np.array(X)
y = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression().fit(X_train, y_train)
regressor.score(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor_2 = RandomForestRegressor(n_estimators=300).fit(X_train, y_train)
regressor_2.score(X_train, y_train)
```

```python
y_pred = regressor_2.predict(X_test)
y_pred = pd.Series(y_pred)
print(y_pred)
```
