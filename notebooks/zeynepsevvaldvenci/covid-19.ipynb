```python
import pandas as pd
covid19 = pd.read_csv("../input/corona-virus-report/covid_19_clean_complete.csv")
```

```python
covid19["Country/Region"]
```

```python
countryList = []
for country in covid19["Country/Region"]:
    for unique in country.split(","):
        if unique not in countryList:
            countryList.append(unique)
print(countryList)
```

```python
countryTotal = {}
countryCount = {} 

for country in countryList:
    countryTotal[country] = 0
    countryCount[country] = 0
print(countryTotal)
print(countryCount)
```

```python
for idx, countries in enumerate(covid19["Country/Region"]):
    
    for country in countries.split(","):
        countryTotal[country] += covid19["Deaths"][idx] 
        countryCount[country] += 1
print(countryTotal)
print("******")
print(countryCount)
```

```python
countryAverage = {}

for country in countryTotal.keys():
    countryAverage[country] = countryTotal[country] / countryCount[country]
print(countryAverage)
```

```python
resultCountry = "temp"
resultDeaths = 0
for country in countryAverage.keys():
    if countryAverage[country] > resultDeaths:
        resultCountry = country
        resultDeaths = countryAverage[country]
        
print("MOST DEATH RATE IS IN " + resultCountry + " AND RATE IS " + str(round(resultDeaths,2)) + " BECAUSE OF CORONA VIRUS.")
```
