```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
data['Geography'] = encoder.fit_transform(data['Geography'])
```

```python
data
```

```python
data['Gender'] = encoder.fit_transform(data['Gender'])
```

```python
data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
data[['CreditScore', 'Age', 'Balance', 'EstimatedSalary']] = scaler.fit_transform(data[['CreditScore', 'Age', 'Balance', 'EstimatedSalary']])
```

```python
data
```

```python
x= data.iloc[:, 0:-1]
y = data.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
from sklearn.linear_model import Perceptron
clf = Perceptron(penalty='l1', alpha=0.1, max_iter=1000)
```

```python
clf.fit(xtrain, ytrain)
```

```python
y_pred = clf.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(ytest, y_pred)
accuracy
```

```python
param_dist = {
    'penalty':['l1', 'l2', 'elasticnet'],
    'alpha':[0.1, 0.001, 0.0001],
    'max_iter':[1000, 10000, 100000]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid = GridSearchCV(clf, param_grid = param_dist, cv = 100, n_jobs = -1)
```

```python
grid.fit(xtrain, ytrain)
```

```python
grid.best_score_
```
