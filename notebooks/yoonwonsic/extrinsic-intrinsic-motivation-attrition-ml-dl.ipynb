# Ⅰ. Preface

## 1. Meaning of Analysis and Direction

 Attrition in company cause cost because the company have to recruit, screen, train and educate for new employee. Additionaly, if not recriut for a long term, remain employee can feel burden. Also, it has a bad influence on group cohesiveness.<br/>
 However, there can be a lot of factors which affect to attrition, and its make managing the problem complex. So it is important that keep appropriate attrition, and analysis cause of attrition is requested.

 Inevitably, I will research about **demographic characteristics** as personel factors. For example, person who prefer stability generally less frequently change the job, so attrition will decrease too. So we cannot ignore influence of personal characteristics to attrition.
 
 Next part is analysis about **job and vocational behavior characteristics**. It can represent employee's awareness and attitude to work.<br/>
 For instance, if certain department's attirtion occur much more frequently than other's, company have to focus on the department. And it is the same to job role or job level. <br/>Also, it will be interesting to search about vocational behavior characteristic. We can expect that someone who have worked at company for a long time, would not resign esasily.
 
Above all things, I will focus on "Motivation".  In this analysis, especially I applicate concept of "**extrinsic reward**" and "**intrinsic motivation**". In other words, if that two factors are not met employee's needs, the employ will leave the company. <br/>
Extrinsic reward is so-called economical/monetary compensation, for example, income and stock option. Naturally, every people always wants more compenstion.<br/>
But human seek more valuable things, such as meaningful role at their work performence. So there are many theories to explain employee's behavior in the company. For instance, Herzberg's two-factor theory insist that there are motivator and hygiene factor. Hygine factors like salary and benefit just can suppres dissatisfaction, while motivators like worthwhile job can lead positive behavior for organization. <br/>
With these basis, I will analyze the effect of compensation on the Attrition, and the effect of subjective factor on the Attrition.

## 2. Set Independant Variables

The data, "WA_Fn-UseC_-HR-Employee-Attrition.csv", which is target of this analysis, has a 35 variables(columns). Among that, I pick 'Attrition' as dependant variable. And pick independant varables and set hypothesis like below.

### 2. (1) Personal(Demographic) Characteristics

#### ① Age
- If someone getting more age, the employee would be harder to seek a great change in life.

#### ② Gender
- It is expected that Gender itself would not affect to attrition.<br/>
  But as a basic demographic varables, it is need to analyze.

#### ③ Marital Status
- Marital status combine with support family. For employee, it is one of main cosideration on leave company.

#### ④ Education
- According to labor economics, high level educated group can face to larger labor market.<br/>
  So they can choose other companies and decide to resign easily.
  
#### ⑤ Distance from Home
- Long commute time bother employees so much. It can stimulate wants to leave company

### 2. (2) Job·Vocational behavior Characteristic

#### ① Department
- If attrition occur at one department much more than others, it is requested that specialized solution

#### ② Job Role / Job Level
- Accoring to 「Job Characteristics theory」(Hackman,  Oldham),ⅰ)Skill Variety, ⅱ)Task Identity, ⅲ)Task Significance, ⅳ)Autonomy, ⅴ)Feedback can reduce absenteeism and turnover. And the factors are related to each job itself.

#### ③ Working Years At Company
- Someone who worked long time at company, can be expected remain.

#### ④ Number of Worked Company, ⑤ Total Working Years, and ⑥  Average Working Years at Each Company
- Employee's tendency to turnover frequently in past, would be valid current,too.
- 'Average Working Years at Each Company' is a derived variable, which is Total Working Years divided by  Number of Worked Company

### 2. (3) Extrinsic Reward

#### ① Monthly Income
- Without doubt, income is powerful method to compensate for employee.
- If emplyoee resign company that offer large income, the employee have to face to big opportunity cost.

#### ② Stock Option Level
- Company can provide the right of choice to buy stock at fixed price.
- It makes relationship between company's growth and employee's long-term profit.

#### ③ Percent of Salary Hike
- Increase of income make employee expect larger income of future.

#### ④ Performance Rating
- Result of performance appraisal means degree of acknowledgment about contribution in the company.

#### ⑤ Years Since Last Promotion
- Personnel congestion is structural problem that makes employees are frustrated at the company

### 2. (4) Intrinsic Motivation
The followings are factor that can prevent empolyee's resignation.<br/>
These are also related with 「Job Embeddedness」(links, fit and sacrifice).

#### ① Job Involvement, ② Job Satisfaction
- These are typical indicators which correspond to intrinsic motivator.

#### ③ Relationship Satisfaction
- According to 「Human Relation theory」(Mayo, Roethlisberger), sometimes relationship between employees is much more important than monetary factors. (It is discoveried in the Hawthore factory study)

#### ④ Environment Satisfaction
- Well built environment make employee involve their company and job

#### ⑤ Traing Times
- To satisfy employee's need for growth, company should offer sufficient developement program.

#### ⑥ Work-Life Balence
- If company request to much time, employee's family will not agree that the employee work at the company continuously.

#### ⑦ Years In Current Role
- If one person repeat the same work for a long time, it should make the person burn out.

### ② Data Conservation for Logistic Regression
It is discrete variable. So, for the analysis, the **Logistic Regression** can be use.<br/>
Therefore the occurrence of event "Attrition" should be converted "1", the other is "0".

```python
dataframe["Attrition_dummy"] = dataframe["Attrition"]
dataframe.loc[dataframe["Attrition"]=="Yes", ["Attrition_dummy"]] = 1
dataframe.loc[dataframe["Attrition"]=="No", ["Attrition_dummy"]] = 0
dataframe["Attrition_dummy"] = dataframe["Attrition_dummy"].astype(int)
```

## Ⅲ. 2. Effect of Demographic Characteristic on the Attrition

### Ⅲ. 2. (1) Age

#### ① Exploratory Data Analysis

```python
age_label = ["~19", "20~24", "25~29", "30~34", "35~39", "40~44", "45~49", "50~54", "55~59"]
age_cate = pd.cut(x=dataframe['Age'], bins=[18, 20, 25, 30, 35, 40, 45, 50, 55, 60], labels=age_label)
dataframe['age_cate'] = age_cate
age_frame = dataframe['age_cate'].value_counts().to_frame().sort_index()
age_frame.index.name = "age"
age_frame["num"] = age_frame["age_cate"]
del age_frame["age_cate"]
age_frame
```

```python
dataframe["Age"].hist(bins=9, grid=False)
plt.rcParams['font.size'] = 12
```

#### ② Visualization Analysis

```python
plt.rcParams['font.size'] = 12
sns.boxplot(x="Attrition", y="Age", data=dataframe)
```

#### ③ Logistic Regression

```python
feature_Age = dataframe[['Age']]
attrition = dataframe['Attrition_dummy']

train_feature_Age, test_feature_Age, train_labels, test_labels = train_test_split(feature_Age, attrition)

scaler = StandardScaler()
train_feature_Age = scaler.fit_transform(train_feature_Age)
test_feature_Age = scaler.transform(test_feature_Age)

model = LogisticRegression()

model.fit(train_feature_Age, train_labels)

print("Train Accuracy : ", model.score(train_feature_Age, train_labels).round(3))
print("Test Accuracy : ", model.score(test_feature_Age, test_labels).round(3))
print("Coefficient : ", model.coef_[0][0].round(3))
```

#### ④ Sub-Conclusion

There is no problem of model accuracy, and overfitting, underfitting, too. <br/>
And the absolute value of coefficient is sufficient level.<br/>
So, I **should pick** this variable to final analysis.

## Ⅳ. 2. Logistic Regression

```python
features = dataframe[["Age", "Marital_Single_dummy", "JobLevel", "AvgWorkingYears", "MonthlyIncome", "StockOptionLevel",
                     "JobInvolvement", "JobSatisfaction", "EnvironmentSatisfaction", "YearsInCurrentRole"]]
attrition = dataframe['Attrition_dummy']

train_features, test_features, train_labels, test_labels = train_test_split(features, attrition)

scaler = StandardScaler()
train_features = scaler.fit_transform(train_features)
test_features = scaler.transform(test_features)

model = LogisticRegression()

model.fit(train_features, train_labels)

print("① Train Accuracy : ", model.score(train_features, train_labels).round(3))
print("② Test Accuracy : ", model.score(test_features, test_labels).round(3))
print("③ Coefficients : ", model.coef_.round(3)[0])
print("④ Confusion Matrix")
print("   ", confusion_matrix(test_labels, model.predict(test_features))[0][0], "  ",
      confusion_matrix(test_labels, model.predict(test_features))[0][1])
print("   ", confusion_matrix(test_labels, model.predict(test_features))[1][0], "  ",
      confusion_matrix(test_labels, model.predict(test_features))[1][0])
print("⑤ Classification Report")
print(classification_report(test_labels, model.predict(test_features)))
```

## Ⅳ. 3. Decision Tree

```python
model2 = DecisionTreeClassifier(max_depth=3)

model2.fit(train_features, train_labels)

print("① Train Accuracy : ", model2.score(train_features, train_labels).round(3))
print("② Test Accuracy : ", model2.score(test_features, test_labels).round(3))
print("③ Confusion Matrix")
print("   ", confusion_matrix(test_labels, model2.predict(test_features))[0][0], "  ",
      confusion_matrix(test_labels, model2.predict(test_features))[0][1])
print("   ", confusion_matrix(test_labels, model2.predict(test_features))[1][0], "  ",
      confusion_matrix(test_labels, model2.predict(test_features))[1][0])
print("④ Classification Report")
print(classification_report(test_labels, model2.predict(test_features)))
```

## Ⅳ. 4.  Random Forest

```python
model3 = RandomForestClassifier(n_estimators=100, max_depth=7)

model3.fit(train_features, train_labels)

print("① Train Accuracy : ", model3.score(train_features, train_labels).round(3))
print("② Test Accuracy : ", model3.score(test_features, test_labels).round(3))
print("③ Confusion Matrix",)
print("   ", confusion_matrix(test_labels, model3.predict(test_features))[0][0], "  ",
      confusion_matrix(test_labels, model3.predict(test_features))[0][1])
print("   ", confusion_matrix(test_labels, model3.predict(test_features))[1][0], "  ",
      confusion_matrix(test_labels, model3.predict(test_features))[1][0])
print("④ Classification Report")
print(classification_report(test_labels, model3.predict(test_features)))
```

# Ⅴ. Final Analysis - Deep Learning

On the basis of former analysis, I have made sigmoid model with four hidden layer.

```python
model4 = Sequential()

model4.add(Dense(512, input_dim=features.shape[1], activation="relu"))
model4.add(Dense(512, activation="relu"))
model4.add(Dense(512, activation="relu"))
model4.add(Dense(512, activation="relu"))
model4.add(Dropout(0.2))

model4.add(Dense(1, activation="sigmoid"))

model4.compile(loss = "binary_crossentropy", optimizer=Adam(lr=0.01), metrics=["accuracy"])

print("① Train Accuracy : ")
model4.fit(train_features, train_labels, batch_size=50, epochs=20, verbose=1)
print("")
print("② Test Accuracy : ")
model4.evaluate(test_features, test_labels)
```
