### Part 2 Data Training, Testing and Making Predictions.

Let's Select Data on the Basis of which we will made predictions.

```python
X_Features = weatherData[['Humidity']]
```

```python
X_Features
```

Now let's select the Data about which predictions will be made.

```python
Y_Prediction = weatherData[['Temperature (C)']]
```

```python
Y_Prediction
```

Now Let's Train and Test the Data.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_Features, Y_Prediction, test_size=0.40, random_state=101)
```

Now let's import the linear regression model in order to train it.

```python
from sklearn.linear_model import LinearRegression
```

Let's initiate the Linear Regression Model.

```python
linearRegressionInstance = LinearRegression()
```

```python
linearRegressionInstance
```

Let's Train the Model.

```python
linearRegressionInstance.fit(X_train,y_train)
```

Now let's check the Intercept.

```python
linearRegressionInstance.intercept_
```

Now let's check the Co-Efficients.

```python
linearRegressionInstance.coef_
```

let's define the coefficient in a describeable form.

```python
coefficientInformation = pandasInstance.DataFrame(data=linearRegressionInstance.coef_,index=X_train.columns,columns=['Co-Efficient Values'])
```

```python
coefficientInformation
```

Now let's Predict the Temperature Values.

```python
temperaturePredictions = linearRegressionInstance.predict(X_test)
```

```python
temperaturePredictions
```

Let's Check the Difference between actual and predicted values in a scatter graph.

```python
matplotlibInstace.scatter(y_test,temperaturePredictions)
```

Now let's check the Distributations of the Residuals.

```python
seabornInstace.distplot(y_test-temperaturePredictions)
```

###### as from Distributation Graph above we can conclude that there is nearly normal distributation.

Let's Do somework for the Errors.

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,temperaturePredictions)
```

```python
metrics.mean_squared_error(y_test,temperaturePredictions)
```

```python
numpyInstace.sqrt(metrics.mean_squared_error(y_test,temperaturePredictions))
```
