```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
file_path = '../input/videogamesales/vgsales.csv'
df = pd.read_csv(file_path)
df.head()
```

```python
plt.figure(figsize=(10,6))

# Add title
plt.title("Average Sales by Years")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.lineplot(x=df['Year'], y=df['Global_Sales'], label = 'Global_Sales')
sns.lineplot(x=df['Year'], y=df['Other_Sales'], label = 'Other_Sales')
sns.lineplot(x=df['Year'], y=df['JP_Sales'], label = 'JP_Sales')
sns.lineplot(x=df['Year'], y=df['EU_Sales'], label = 'EU_Sales')
# Add label for vertical axis
plt.ylabel("Sales")
```
