```python
label_encoder = LabelEncoder()
for col in train.columns[train.dtypes == "object"]:
    train[col] = label_encoder.fit_transform(train[col].astype('str'))

for col in test.columns[test.dtypes == "object"]:
    test[col] = label_encoder.fit_transform(test[col].astype('str'))
```

```python
import random
male = train[train['gender'] == 0]
female = train[train['gender'] == 1]

## empty list for storing mean sample
m_mean_samples = []
f_mean_samples = []

for i in range(50):
    m_mean_samples.append(np.mean(random.sample(list(male['math score']),50,)))
    f_mean_samples.append(np.mean(random.sample(list(female['math score']),50,)))
    
ttest,pval = stats.ttest_rel(m_mean_samples, f_mean_samples)
print(float(pval))
if pval<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

```python
import random
grp0= train[train['race/ethnicity']==0]
grp1= train[train['race/ethnicity']==1]
grp2= train[train['race/ethnicity']==2]

p0_mean_samples=[]
p1_mean_samples=[]
p2_mean_samples=[]

for i in range(50):
    p0_mean_samples.append(np.mean(random.sample(list(grp0['math score']),50,)))
    p1_mean_samples.append(np.mean(random.sample(list(grp1['math score']),50,)))
    p2_mean_samples.append(np.mean(random.sample(list(grp2['math score']),50,)))
  
                          
F, p = stats.f_oneway(p0_mean_samples, p1_mean_samples, p2_mean_samples)
print("p-value for significance is: ", p)
if p<0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```
