### So there are many columns which are not correlated which each other. So we'll just check the correlation of our independent variable and target variable.

```python
summary['MinTemp'].corr(summary['MaxTemp'])
```

### They are highly correlated. That means we can apply Machine learning algorithm to it.

Importing ML Libraries

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

### Splitting the variables into train and test set.

```python
xtrain,xtest,ytrain,ytest = train_test_split(summary[['MinTemp']],summary[['MaxTemp']],
                                             test_size = 0.30,
                                            random_state = 123)
```

```python
xtrain.size
```

```python
xtest.size
```

### Creating Regression object

```python
lr = LinearRegression()
```

### Fitting the training sets into the object

```python
lr.fit(xtrain, ytrain)
```

### Predicting the test set based on training set

```python
ypred = lr.predict(xtest)
```

```python
ypred
```

## Visualizing the training set with the Linear Regression line

```python
plt.scatter(xtrain,ytrain, color='red')
plt.plot(xtrain,lr.predict(xtrain))
plt.xlabel('Minimum Temperature')
plt.ylabel('Maximum Temperature')
plt.title('Minimum Vs Maximum Temperature')
plt.show()
```

## Visualizing the test set with the Linear Regression line

```python
plt.scatter(xtest,ytest, color='red')
plt.plot(xtest,lr.predict(xtest))
plt.xlabel('Minimum Temperature')
plt.ylabel('Maximum Temperature')
plt.title('Minimum Vs Maximum Temperature')
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
r2_score(ytest,ypred)
```

### So the accuracy of the model is nearly 77%

```python
np.sqrt(mean_squared_error(ytest,ypred)) 
```

# Conclusion - The above value represents that our predicted temperature is 4.13 degrees away from the actual temperature.
