```python
churn.drop('customerID',axis=1,inplace=True)
```

```python
unique_values=[]
for i in churn.columns:
    unique_values.append((i,churn[i].unique()))
unique_values
```

```python
churn['gender']=churn['gender'].map({'Female':0,'Male':1})
churn['Dependents']=churn['Dependents'].map({'Yes':1,'No':0})
churn['Partner']=churn['Partner'].map({'Yes':1,'No':0})
churn['PhoneService']=churn['PhoneService'].map({'Yes':1,'No':0})
churn['MultipleLines']=churn['MultipleLines'].map({'No phone service':0,'Yes':1,'No':0})
churn['InternetService']=churn['InternetService'].map({'DSL':1,'Fiber optic':1,'No':0})
churn['OnlineSecurity']=churn['OnlineSecurity'].map({'No':0, 'Yes':1, 'No internet service':0})
churn['OnlineBackup']=churn['OnlineBackup'].map({'Yes':1, 'No':0, 'No internet service':0})
churn['DeviceProtection']=churn['DeviceProtection'].map({'No':0, 'Yes':1, 'No internet service':0})
```

```python
churn['TechSupport']=churn['TechSupport'].map({'No':0, 'Yes':1, 'No internet service':0})
churn['StreamingTV']=churn['StreamingTV'].map({'No':0, 'Yes':1, 'No internet service':0})
churn['StreamingMovies']=churn['StreamingMovies'].map({'No':0, 'Yes':1, 'No internet service':0})
churn['Contract']=churn['Contract'].map({'Month-to-month':1, 'One year':2, 'Two year':3})
churn['PaperlessBilling']=churn['PaperlessBilling'].map({'Yes':1,'No':0})
churn['PaymentMethod']=churn['PaymentMethod'].map({'Electronic check':1, 'Mailed check':2, 'Bank transfer (automatic)':3,
         'Credit card (automatic)':4})
churn['Churn']=churn['Churn'].map({'Yes':1,'No':0})
```

```python
unique_values=[]
for i in churn.columns:
    unique_values.append((i,churn[i].unique()))
unique_values
```

```python
churn.dtypes
```

```python
churn['TotalCharges']=pd.to_numeric(churn['TotalCharges'], errors="coerce")
```

```python
churn.drop(churn[churn['TotalCharges'].isnull()].index, inplace=True)
churn.reset_index(drop=True, inplace=True)
```

```python
X=churn.iloc[:,:-1]
Y=churn['Churn']
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
lgr=LogisticRegression()
knn=KNeighborsClassifier(n_neighbors=9, leaf_size=20)
dt=DecisionTreeClassifier(max_depth=4, random_state=42)
rf=RandomForestClassifier(n_estimators=300,min_samples_leaf=0.16, random_state=42)

lgr.fit(X_train,y_train)
pred_lgr=lgr.predict(X_test)
print('Logistic Regression')
print('Confusion Matrix:\n',confusion_matrix(y_test,pred_lgr))
print('Classification Report:\n',classification_report(y_test,pred_lgr))
print('\n')
print('K Neighbors Classifier')
knn.fit(X_train,y_train)
pred_knn=knn.predict(X_test)
print('Confusion Matrix:\n',confusion_matrix(y_test,pred_knn))
print('Classification Report:\n',classification_report(y_test,pred_knn))
print('\n')
print('Decision Tree Classifier')
dt.fit(X_train,y_train)
pred_dt=dt.predict(X_test)
print('Confusion Matrix:\n',confusion_matrix(y_test,pred_dt))
print('Classification Report:\n',classification_report(y_test,pred_dt))
print('\n')
print('Random forest Classifier')
rf.fit(X_train,y_train)
pred_rf=rf.predict(X_test)
print('Confusion Matrix:\n',confusion_matrix(y_test,pred_rf))
print('Classification Report:\n',classification_report(y_test,pred_rf))
```
