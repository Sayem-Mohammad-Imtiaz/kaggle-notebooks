```python
# seperate part of the title before (:) from the other parts and put it in a new column

df['main title'] = df['title'].apply(lambda x:x.split(':')[0])
df.head()
```

```python
sns.countplot(data=df, x='main title' )
```

```python
# plot the three more titles
df_new = df['title'].value_counts().sort_values(ascending= False).head(3)
df_new = df_new.reset_index()
plt.figure(figsize=(9,5) , dpi = 100)
sns.barplot(x =df_new['index'] ,y= df_new['title'] , data = df_new)
plt.xlabel('title')
plt.ylabel('count')

```

```python
df['timeStamp'].iloc[0]
```

```python
# change the type of column timeStamp to datetime
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
type(df['timeStamp'].iloc[0])
```

```python
df['hour'] = df['timeStamp'].apply(lambda time : time.hour)
```

```python
df['day of week'] = df['timeStamp'].apply(lambda time:time.dayofweek)
```

```python
df.head()
```

```python
days = ['Monday' , 'Thursday' , 'Wendsday' , 'Tuesday' , 'Friday' , 'Saturday' , 'Sunday']
df['day of week'] = df['day of week'].apply(lambda x:days[int(x)])
df.head()
```

```python
plt.figure(dpi = 100)
sns.countplot(data=df , x = 'day of week', hue = 'main title')
plt.legend(loc = (1.1,0.7))
```
