1. Date is not in correct datatype , convert Date column from string to datetime datatype .

```python
import datetime
```

The dates are present in the following format : 07-01-2014  i.e Day-Month-Year . format='%d-%m-%Y'

```python
df['Date'] = pd.to_datetime(df['Date'] , format='%d-%m-%Y')
```

```python
df.Price.unique()
```

For Price Column :
1. replace the value "Free" to 0.
2. remove rupee sign '₹ 64.00' .
3. remove .00 paisa from the end  '₹ 209.00' .
4. remove commas from the price '₹ 5,449.00' .
4. change the datatype from string to integer .

```python
df['Price'] = df.Price.replace('Free', 0)
```

```python
df.Price.unique()
```

1. "Free" is replaced by 0 .

```python
for i in range (0 , df.shape[0]):
    r = df['Price'][i]
#     print(r)
    if (r !=0):
#         print(r[2:])
        df['Price'][i] = r[2:-3]
    else:
        pass
```

```python
df['Price'].unique()
```

1. Rupee sign is removed .
2. .00 is removed .

```python
for i in range (0 , df.shape[0]):
    r = df['Price'][i]
#     print(r)
    if (r !=0):
        df['Price'][i] = df['Price'][i].replace(',' , '')
```

```python
df['Price'].unique()
```

1. "," commas are also been removed .

```python
df['Price'] = df['Price'].astype(int)
```
