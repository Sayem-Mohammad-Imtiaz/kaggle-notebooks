```python
import numpy as np
import pandas as pd
from sklearn import tree

from sklearn import tree
#import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
```

**Reading Dataset**

```python
df=pd.read_csv("../input/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv")
df
```

```python
df=df.iloc[:20,:]
df
```

```python
plt.scatter(df.tenure,df.Churn,marker='+',color='red') #Scatter plotting
```

```python
df.shape
```

```python
#Split rows into train_test_split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['tenure']],df.Churn, test_size=0.33)
X_train #Data samples for training our model
```

```python
X_test #Data samples for testing our model
```

```python
from sklearn.linear_model import LogisticRegression #Import logistic regression 
model = LogisticRegression()
model.fit(X_train,y_train) #fit function is for training model

model.predict(X_test) #Prediction
```

```python
model.score(X_test,y_test) #show/checking accuracy of model
```

```python
model.predict_proba(X_test) #Prediction probability, probability of sample of one class with other class (array( [notchurn(no)], [churn(yes)] ))
```
