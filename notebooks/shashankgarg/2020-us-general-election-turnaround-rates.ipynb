```python
df = pd.read_csv('../input/2020-us-general-election-turnout-rates/2020 November General Election - Turnout Rates.csv')
```

# **Data Cleaning**

```python
df.describe()
```

```python
df.dtypes
```

```python
df.columns
```

```python
transform_columns = {
    'State': 'category',
    'Official/Unofficial': 'category',
    'State Abv': 'category'
}
df['Total Ballots Counted (Estimate)'] = df['Total Ballots Counted (Estimate)'].str.split(',').str.join('').astype(int)
df['Vote for Highest Office (President)']= df['Vote for Highest Office (President)'].str.split(',').str.join('').astype(float)
df['Voting-Eligible Population (VEP)'] = df['Voting-Eligible Population (VEP)'].str.split(',').str.join('').astype(int)
df['Voting-Age Population (VAP)'] = df['Voting-Age Population (VAP)'].str.split(',').str.join('').astype(int)
df['VEP Turnout Rate']= df['VEP Turnout Rate'].str.split('%').str.join('').astype(float)
df['% Non-citizen']= df['% Non-citizen'].str.split('%').str.join('').astype(float)
df['Prison'] = df['Prison'].str.split(',').str.join('').astype(int)
df['Probation'] = df['Probation'].str.split(',').str.join('').astype(int)
df['Parole'] = df['Parole'].str.split(',').str.join('').astype(int)
df['Total Ineligible Felon'] = df['Total Ineligible Felon'].str.split(',').str.join('').astype(int)
df = df.astype(transform_columns)
df = df.drop(columns=['Source', 'Overseas Eligible'])
```
