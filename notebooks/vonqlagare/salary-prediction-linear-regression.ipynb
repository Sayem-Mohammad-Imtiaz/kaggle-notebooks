```python
# Initializing all the needed libraries
import matplotlib.pyplot as plt
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
```

```python
# Creating dataFrame using pandas
df = pd.read_csv("../input/salary-data-simple-linear-regression/Salary_Data.csv")
df.head()
```

```python
# Checking if the data is applicable for linear regression

plt.scatter(df[['YearsExperience']], df.Salary)
plt.grid()
```

```python
# Getting information about the columns
df.info()
```

```python
# Looking for outliers in the data

sns.boxplot(x=df.Salary, data=df)
```

```python
# Splitting the data into test amd train

from sklearn.model_selection import train_test_split
```

```python
# slitting x and y from the dataFrame

x = df.YearsExperience
y = df.Salary
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
# Creatiing a model using linear regression

from sklearn.linear_model import LinearRegression
```

```python
# increasing the demension of test and train data

x_train = x_train[:, np.newaxis]
x_test = x_test[:, np.newaxis]
```

```python
# Initializing a linear regression object

reg = LinearRegression()
```

```python
# Training the model using the split data sets

reg.fit(x_train, y_train)
```

```python
# Predicting the salary using test data

pred = reg.predict(x_test)
```

```python
# Plotting the result of prediction

plt.title("Salary Prediction")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")

plt.scatter(x_test, y_test)
plt.plot(x_test, pred)
```
