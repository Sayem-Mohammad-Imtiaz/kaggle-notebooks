```python
indexData = pd.read_csv('../input/stock-exchange-data/indexData.csv')
indexInfo = pd.read_csv('../input/stock-exchange-data/indexInfo.csv')
indexProcessed = pd.read_csv('../input/stock-exchange-data/indexProcessed.csv')
```

```python
def adjustedCurrency(df):
    conversion_dict = {"NYA" : 1.0, "IXIC" : 1.0, "HSI" : 0.13, "000001.SS" : 0.15, "N225" : 0.0091, "N100" : 1.19, "399001.SZ" : 0.15,"GSPTSE" : 0.80, "NSEI" : 0.014, "GDAXI" : 1.19, "KS11" : 0.00087, "SSMI" : 1.09, "TWII" : 0.036, "J203.JO" : 0.070 }
    temp = []
    for row in df.itertuples():
        temp.append(conversion_dict[row[1]])
    df["Open"] = df["Open"]*temp
    df["High"] = df["High"]*temp
    df["Low"] = df["Low"]*temp
    df["Close"] = df["Close"]*temp
    df["Adj Close"] = df["Adj Close"]*temp
```
