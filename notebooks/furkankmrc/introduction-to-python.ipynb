# **Loop Data Structures (while and for)**

```python
i=0
while i!=5 :
    print('i is: ',i)
    i+=1
print (i,' is equal to 5')
```

```python
#stay in loop if conditions is true
lis =[1,2,3,4,5]
for i in lis:
    print('i is: ',i)
print('')

# Enumerate index and value of list
for index,value in enumerate(lis):
    print(index,":",value)
print('')

#for dictionaries
dictionary={'spain' : 'madrid' , 'france' : 'paris' , 'Turkey' : 'Ankara'}
for key,value in dictionary.items():
    print(key,":",value)
print('')

for index,value in data[['Attack']][0:10].iterrows():
    print(index," : ",value)
```
