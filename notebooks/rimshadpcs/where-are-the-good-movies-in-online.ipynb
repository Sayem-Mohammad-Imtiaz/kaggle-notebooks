## Observations

```python
movies_Genre=data.copy()
del movies_Genre['Genres']
s = data['Genres'].str.split(',').apply(pd.Series, 1).stack()
s.index = s.index.droplevel(-1)
s.name = 'Genres'
movies_Genre = movies_Genre.join(s)
```

```python
# to seperate genre column in dataset
seperated_genres = data['Genres'].str.get_dummies(',')

# to concatenate two dataframes
data = pd.concat([data, seperated_genres], axis = 1, sort = False)
```

```python
# Missing values
data.isnull().sum()
```

```python
# dropping columns with large missing values
data.drop(['Rotten Tomatoes'],axis=1,inplace=True)
```

## Platform wise content analysis

```python
# Number of movies in each platforms
netflix_count = len(data[data['Netflix']== 1].index)
hulu_count = len(data[data['Hulu']== 1].index)
prime_video_count = len(data[data['Prime Video']==1].index)
disney_count = len(data[data['Disney+']==1].index)
print('Netflix - ',netflix_count)
print('Hulu - ',hulu_count)
print('Prime Video - ',prime_video_count)
print('Disney+ - ', disney_count)
```

* Prime video leads in terms of number of movies.
* Netflix is second
* Prime video has thrice more movies than Netflix
* Movies in Hulu and Disney+ are less than 1000

```python
# Saving this to a data frame

label=['Netflix','Hulu', 'Prime Video','Disney+']
count=[netflix_count,hulu_count,prime_video_count,disney_count]
platform = pd.DataFrame(
    {'Platform': label,
     'Count': count,
    })
platform
```

```python
import plotly.express as px
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
fig = px.pie(platform,names='Platform', values='Count')
fig.update_traces(rotation=45, pull=[0.1,0.03,0.03,0.03,0.03],textinfo="percent+label", title='Number of movies per platform')
fig.show()
```

## Imdb rating wise analysis

```python
data['Genres'].value_counts().head(10)
```

                                            Imdb rating wise analysis with respect to Genre

```python
top_5_genres = ['Drama','Documentary','Comedy', 'Comedy,Drama','Horror','Drama,Romance','Comedy,Drama,Romance',
            'Horror,Thriller','Comedy,Romance','Drama,Thriller']
table = data.loc[:,['Year','Genres','IMDb']]
table['AvgRating'] = table.groupby([table.Genres,table.Year])['IMDb'].transform('mean')
table.drop('IMDb', axis=1, inplace=True)
table = table[(table.Year>1999) & (table.Year<2021)]
table = table.loc[table['Genres'].isin(top_5_genres)]
table = table.sort_values('Year')
```

```python
fig=px.bar(table,x='Genres', y='AvgRating', animation_frame='Year', 
           animation_group='Genres', color='Genres', hover_name='Genres', range_y=[0,10])
fig.update_layout(showlegend=False)
fig.show()
```

* "Documentary" shows mostly had an edge on other genres from 2000 to 2020 by rating wise
* The only period Another genre overtook "Documentary" was in 2010, by "Drama,Romance" with "7.375" as average rating
* Throuh out the 2000-2020 periods "Drama,Romance", "Drama" "Comedy,Drama,Romance" are the only genres come close with    "Documentary" genre
* It has found "Horror" and "Horror,Thriller" are the least favourites in rating wise analysis with their graphs never overtook other top 8 genres at all.
* The average rating of "Horror, thriller" genre was 2.5 in 2020 which is the lowest in the period  of 2000-2020 followed by "Horror"'s 2.55 in 2010
