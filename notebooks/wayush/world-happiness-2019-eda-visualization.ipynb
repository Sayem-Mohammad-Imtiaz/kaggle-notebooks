# EDA: Visualization with matplotlib and seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
import seaborn as sns
```

# 1. Load the dataset into a Pandas dataframe

```python
# Reading Dataset
dataSet = pd.read_csv("../input/world-happiness/2019.csv")
```

```python
# Rounding Upto 2 Decimal Point
pd.set_option('precision',2)
```

```python
#Changing the names of columns header
dataSet.columns = ['Rank','Country','Score','GdpPerCapita' ,'SocialSupport','Healthy life expectancy','Freedom to make life choices' ,'Generosity', 'Perceptions of corruption']
```

# 4. Create one or more histograms

```python
# Creating a bar graph for top 5 ranking countries with respect to Genoriosity

plt.bar(dataSet.head(5)['Country'],dataSet.head(5)['Generosity'])
plt.ylabel("Genoriosity", fontsize=15,color="Green")
plt.xlabel("Country", fontsize=15,color="Blue")
```

```python
# Creating a bar graph for top 5 ranking countries with respect to Social Support

plt.bar(dataSet.head(5)['Country'],dataSet.head(5)['SocialSupport'])
plt.ylabel("SocialSupport", fontsize=15,color="Green")
plt.xlabel("Country", fontsize=15,color="Blue")
```

```python
#Plot a line graph for Top 20 Entries ,comparing GDP per Capita with Health Life Expectancy

sns.lineplot(data = dataSet.head(20), x = "GdpPerCapita", y = "Healthy life expectancy")
```

```python
sns.pairplot(dataSet, vars = ['Rank','Score','GdpPerCapita' ,'SocialSupport','Healthy life expectancy'], kind = 'kde',corner = True) 
```

```python
#Plot a line graph for Top 20 Entries ,comparing Social support with Health Life Expectancy

sns.lineplot(data = dataSet.head(20), x = "SocialSupport", y = "Healthy life expectancy")
```
