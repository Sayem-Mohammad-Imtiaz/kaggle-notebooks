```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

from sklearn.decomposition import PCA
from sklearn.preprocessing import RobustScaler,StandardScaler,MinMaxScaler
from sklearn.pipeline import make_pipeline

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../input/pokemon/pokemon.csv')
df = data.copy()
pd.set_option('display.max_row',df.shape[0])
pd.set_option('display.max_column',df.shape[1]) 
df.head()
```

```python
corr = df.corr(method='pearson').abs()

fig = plt.figure(figsize=(20,12))
ax = sns.heatmap(corr, annot=True, cmap='Blues', vmin=0, vmax=+1)

ax.axhline(18,0,18/34, ls='--', linewidth=3, color="red")
ax.axvline(18-0.05,16/34,34/34, ls='--', linewidth=3, color="red")
ax.axhline(0+0.05,0,18/34, ls='--', linewidth=3, color="red")
ax.axvline(0.05,16/34,34/34, ls='--', linewidth=3, color="red")

ax.axhline(18,18/34,34/34, ls='--', linewidth=3, color="green")
ax.axvline(18+0.05,0,16/34, ls='--', linewidth=3, color="green")
ax.axhline(34-0.07,18/34,34/34, ls='--', linewidth=3, color="green")
ax.axvline(34-0.07,0,16/34, ls='--', linewidth=3, color="green")

plt.title('Pearson Correlation')
plt.show()
```

## Comments
- We can see some correlations inside of attack types (red) and other general stats (green)
- We can also see that general stats don't seem to affect the efficiency of attacks against specific types (not framed in green or red)

# Visualizing types

```python
preprocessor = make_pipeline(MinMaxScaler())
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numeric_df = df.select_dtypes(include=numerics)
numeric_df = numeric_df.fillna(numeric_df.mean())
PCAPipeline = make_pipeline(preprocessor, PCA(n_components=3,random_state=0))
```

```python
df['type1'].unique()
```

```python
df['type2'].unique()
```

```python
df['type2'] = df['type2'].fillna('None')
```

```python
PCA_df = pd.DataFrame(PCAPipeline.fit_transform(numeric_df))
PCA_df = pd.concat([PCA_df, df['type1']], axis=1)
PCA_df = pd.concat([PCA_df, df['type2']], axis=1)
PCA_df.head()
```

```python
figure1 = px.scatter_3d(PCA_df,
        x=0, 
        y=1, 
        z=2, 
        color = 'type1',
                       width=600, height=800)
figure1.update_traces(marker=dict(size=5,
                              line=dict(width=0.2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))

figure1.show()
```

```python
figure2 = px.scatter_3d(PCA_df,
        x=0, 
        y=1, 
        z=2, 
        color = 'type2',
                       width=600, height=800)

figure2.update_traces(marker=dict(size=5,
                              line=dict(width=0.2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))
figure2.show()
```

# Strength analysis

```python
df=data.copy()
df.head()
```

```python
strength_df = pd.DataFrame()
strength_features = ['against_bug','against_dark','against_dragon','against_electric','against_fairy',
                     'against_fight','against_fire','against_flying','against_ghost','against_grass',
                     'against_ground','against_ice','against_normal','against_poison','against_psychic',
                     'against_rock','against_steel','against_water']
classic_features = ['attack','defense','hp','sp_attack','sp_defense','speed','type1','type2']
strength_df["against_overall"] = df[strength_features].sum(axis=1)
strength_df = pd.concat([df['name'],strength_df,df[classic_features]],axis=1)
strength_df.head()
```

```python
strength_df['attack_overall'] = strength_df['attack']+strength_df['sp_attack']
strength_df['defense_overall'] = strength_df['defense']+strength_df['sp_defense']
strength_df.head()
```

```python
figure3 = px.scatter_3d(strength_df,
                        x='attack_overall', 
                        y='defense_overall', 
                        z='against_overall', 
                        color = 'type1',
                        hover_name  = 'name',
                        width=600,
                        height=800)

figure3.update_traces(marker=dict(size=4,
                              line=dict(width=0.2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))

figure3.show()
```

```python
strongest_df = strength_df[(strength_df['attack_overall']>270) | (strength_df['defense_overall']>290) | (strength_df['against_overall']>23)]
figure4 = px.scatter_3d(strongest_df,
                        x='attack_overall', 
                        y='defense_overall', 
                        z='against_overall', 
                        color = 'type1',
                        hover_name  = 'name',
                        width=600,
                        height=800)

figure4.update_traces(marker=dict(size=4,
                              line=dict(width=0.2,
                                        color='DarkSlateGrey')),
                  selector=dict(mode='markers'))

figure4.show()
```
