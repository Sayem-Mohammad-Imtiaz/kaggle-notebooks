```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("../input/habermans-survival-data-set/haberman.csv",header=None, names=['age', 'year', 'nodes', 'status'])
df.head()
```

```python
df.info()
```

Observation :-> There is no Null Data in this dataframe 

```python
df.describe()
```

Obervation :-> The Age limit is between 30 to 83 . Means all the patients those who are gone for treatment are in the age in between 30 to 83.

```python
df['age'].value_counts()
```

Observation :-> The more affected age limit is in range of 50 to 55

```python
df['status'].value_counts()/len(df)*100
```

Observation :-> 73.50 % of patients got survived after the treatment 

```python
df.corr()
```

Observation :-> From here we can say that nodes are the best way to state the status of the patients.
    we are getting correlation values of nodes and status as 0.28 which is higher than others.

```python
df['status'] =df['status'].map({1:"YES", 2:"NO"})
sns.set_style("whitegrid");
plt.rcParams["figure.figsize"] = (18,8)
sns.FacetGrid(df, hue="status", size=5) \
   .map(plt.scatter, "year", "age") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(df, hue="status", size=8) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(df, hue="status", size=8) \
   .map(sns.distplot, "year") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(df, hue="status", size=8) \
   .map(sns.distplot, "nodes") \
   .add_legend();
plt.show();
```

Observation :-> Based on the above plot we can say that nodes are the based parameters for check the status of patients.
