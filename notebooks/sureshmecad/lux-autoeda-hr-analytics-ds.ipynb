### Titanic : Recommendations based on user intent

```python
df1 = pd.read_csv("/kaggle/input/titanic/train.csv")
df1
```

```python
df1.intent = ['Survived']
df1
```

```python
df1.intent = ['Age', 'Fare']
df1
```

```python
vis = df1.exported
vis
```

```python
# Create individual visualizations with Vis objects
from lux.vis.Vis import Vis
Vis(["Survived"],df1)
```

```python
from lux.vis.Vis import Vis
Vis(["Embarked=C","Fare"],df1)
```

```python
from lux.vis.VisList import VisList
VisList(["Embarked=?","Pclass"],df1)
```
