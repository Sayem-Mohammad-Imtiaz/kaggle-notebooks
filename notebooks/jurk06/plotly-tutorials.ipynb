```python
timesData=pd.read_csv('../input/world-university-rankings/timesData.csv')
```

* Taking the first hundres values from the dataset

```python
df = timesData.iloc[:100,:]
```

```python
df.shape
```

# Plotly 

```python
import plotly as py
import plotly.express as px
```

```python
df.info()
```

# Line Diagram

```python
fig=px.line(x=df['university_name'], y=df['citations'] ,title="University Vs Citation")
print(fig)
fig.show()
```

# Dataset Statistic

```python
df.describe()
```

# Histogram 

```python
fig=px.histogram(df, x="citations",title="Citataions", nbins=45, labels={'x':'citation', 'y':'total_counts'}
                ,opacity=0.8)
fig.show()
```

# stacked Histogram

```python
fig=px.histogram(df, x="teaching",color="country", hover_data=df.columns)
fig.show()
```

```python
fig=px.histogram(df, x="university_name", y="citations", marginal='box',hover_data=df.columns)
fig.show()
```

```python
import plotly.graph_objects as go
```

```python
fig=go.Figure(data=[go.Histogram(x=df['citations'], histnorm='probability')])
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Histogram(x=df['citations']))
fig.add_trace(go.Histogram(x=df['teaching']))
fig.update_layout(barmode='stack')
fig.show()
```

# Bar Chart 

* To plot number of counts
* To plot number of sums

```python
fig=go.Figure()
fig.add_trace(go.Histogram(histfunc="count", x=df['country'], y=df['citations'], name="Counts of Citations"))
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Histogram(histfunc="sum", x=df['country'], y=df['income'], name="Counts of income"))
fig.show()
```

```python
fig=go.Figure()
fig.add_trace(go.Histogram(histfunc="sum", x=df['country'], y=df['international'], name="Sum of international"))
fig.show()
```
