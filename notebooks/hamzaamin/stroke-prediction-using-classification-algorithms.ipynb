<CENTER><h1 STYLE='COLOR:BLUE'> APPLYING CLASSIFIERS ON IMBALANCED DATASET</H1></CENTER>

<h2>SPLITTING DATASET INTO TRAINING AND TESTING</H2>

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,shuffle=True,random_state=30)
```

<H2>1.APPLYING LOGISTIC REGRESSION </H2>

```python
from sklearn.linear_model import LogisticRegression
ilr=LogisticRegression(random_state=30)
ilr.fit(xtrain,ytrain)
ilrpytest=ilr.predict(xtest)
print('accuracy_score\n',accuracy_score(ytest,ilrpytest))
print('precision_score\n',precision_score(ytest,ilrpytest))
print('recall_score\n',recall_score(ytest,ilrpytest))
print('f1_score\n',f1_score(ytest,ilrpytest))
ilr_acc_sc=accuracy_score(ytest,ilrpytest)
ilr_pr_sc=precision_score(ytest,ilrpytest)
ilr_rec_sc=recall_score(ytest,ilrpytest)
ilr_f1_sc=f1_score(ytest,ilrpytest)
```

<h1>2.APPLYING SVM  </H1>

```python
from sklearn.svm import SVC
isvc=SVC(random_state=30)
isvc.fit(xtrain,ytrain)
isvcpytest=isvc.predict(xtest)
print('accuracy_score\n',accuracy_score(ytest,isvcpytest))
print('precision_score\n',precision_score(ytest,isvcpytest))
print('recall_score\n',recall_score(ytest,isvcpytest))
print('f1_score\n',f1_score(ytest,isvcpytest))
isvc_acc_sc=accuracy_score(ytest,isvcpytest)
isvc_pr_sc=precision_score(ytest,isvcpytest)
isvc_rec_sc=recall_score(ytest,isvcpytest)
isvc_f1_sc=f1_score(ytest,isvcpytest)
```

<H2>3.APPLYING DECISION TREE</H2>

```python
from sklearn.tree import DecisionTreeClassifier
idt=DecisionTreeClassifier(random_state=30)
idt.fit(xtrain,ytrain)
idtpytest=idt.predict(xtest)
print('accuracy_score\n',accuracy_score(ytest,idtpytest))
print('precision_score\n',precision_score(ytest,idtpytest))
print('recall_score\n',recall_score(ytest,idtpytest))
print('f1_score\n',f1_score(ytest,idtpytest))
idt_acc_sc=accuracy_score(ytest,idtpytest)
idt_pr_sc=precision_score(ytest,idtpytest)
idt_rec_sc=recall_score(ytest,idtpytest)
idt_f1_sc=f1_score(ytest,idtpytest)
```

<h1>4.NAIVE BAYES</H1>

```python
from sklearn.naive_bayes import GaussianNB
inb=GaussianNB()
inb.fit(xtrain,ytrain)
inbpytest=inb.predict(xtest)
print('accuracy_score\n',accuracy_score(ytest,inbpytest))
print('precision_score\n',precision_score(ytest,inbpytest))
print('recall_score\n',recall_score(ytest,inbpytest))
print('f1_score\n',f1_score(ytest,inbpytest))
inb_acc_sc=accuracy_score(ytest,inbpytest)
inb_pr_sc=precision_score(ytest,inbpytest)
inb_rec_sc=recall_score(ytest,inbpytest)
inb_f1_sc=f1_score(ytest,inbpytest)
```

<H1>5.KNEIGHBOURS CLASSIFIER</H1>

```python
from sklearn.neighbors import KNeighborsClassifier
iknn=KNeighborsClassifier()
iknn.fit(xtrain,ytrain)
iknnpytest=iknn.predict(xtest)
print('accuracy_score\n',accuracy_score(ytest,iknnpytest))
print('precision_score\n',precision_score(ytest,iknnpytest))
print('recall_score\n',recall_score(ytest,iknnpytest))
print('f1_score\n',f1_score(ytest,iknnpytest))
iknn_acc_sc=accuracy_score(ytest,iknnpytest)
iknn_pr_sc=precision_score(ytest,iknnpytest)
iknn_rec_sc=recall_score(ytest,iknnpytest)
iknn_f1_sc=f1_score(ytest,iknnpytest)
```

<h1 style=color:blue>1.Applying grid search cv with Logistic regression</h1>

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score,confusion_matrix
grid_values = {'penalty': ['l1', 'l2'],'C':[0.001,.009,0.01,.09,1,5,10,25],'random_state':list(range(1,100))}
grid_lr_acc = GridSearchCV(lr, param_grid = grid_values)
grid_lr_acc.fit(xtrain, ytrain)
#Predict values based on new parameters
lrgpytest= grid_lr_acc.predict(xtest)
print('Accuracy Score : ' + str(accuracy_score(ytest,lrgpytest)))
print('Precision Score : ' + str(precision_score(ytest,lrgpytest)))
print('Recall Score : ' + str(recall_score(ytest,lrgpytest)))
print('F1 Score : ' + str(f1_score(ytest,lrgpytest)))
```

```python
lr_acc_sc=accuracy_score(ytest,lrgpytest)
lr_pr_sc=precision_score(ytest,lrgpytest)
lr_rec_sc=recall_score(ytest,lrgpytest)
lr_f1_sc=f1_score(ytest,lrgpytest)
```

```python
grid_lr_acc.best_params_
```

```python
grid_lr_acc.best_score_
```
