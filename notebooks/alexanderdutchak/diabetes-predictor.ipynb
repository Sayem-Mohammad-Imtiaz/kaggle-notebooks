```python
data = pd.read_csv('/kaggle/input/pima-indians-diabetes-database/diabetes.csv')

data.head()
```

```python
y = data.Outcome
X = data.drop(['Outcome'], axis=1)

X.head()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.9, 
                                                      test_size=0.1, random_state=0)
```

**RANDOM FOREST**

```python
# Test different models
def model_select(X_train, X_valid, y_train, y_valid):
    test_preds = []
    for n in range(50, 401, 50):
        model = RandomForestRegressor(n_estimators=n, random_state=0)
        model.fit(X_train, y_train)
        preds = model.predict(X_valid)
        test_preds.append(str(n) + ' : ' + str(mean_absolute_error(y_valid, preds)))
    return test_preds
    
model_select(X_train, X_valid, y_train, y_valid)
```

```python
my_model = RandomForestRegressor(n_estimators=250, random_state=0)

my_model.fit(X, y)

test_preds = my_model.predict(X_valid)

print(my_model.score(X_valid, y_valid))
print(mean_absolute_error(y_valid, test_preds))
```

**DECISION TREE**

```python
from sklearn.tree import DecisionTreeClassifier
for i in range(1, 20):
    dtc = DecisionTreeClassifier(max_depth=i, random_state=1)
    dtc = dtc.fit(X_train, y_train)
    print(str(i) + ' : ' + str(dtc.score(X_valid, y_valid)))
```

```python
best_dtc = DecisionTreeClassifier(max_depth=5, random_state=1)
best_dtc.fit(X, y)
dtc_preds = best_dtc.predict(X_valid)
print(best_dtc.score(X_valid, y_valid))
print(mean_absolute_error(y_valid, dtc_preds))
```

**LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr_model = LogisticRegression(random_state=1)
lr_model.fit(X, y)
lr_preds = lr_model.predict(X_valid)
print('ACCURACY: ' + str(100*accuracy_score(y_valid, lr_preds)))
```
