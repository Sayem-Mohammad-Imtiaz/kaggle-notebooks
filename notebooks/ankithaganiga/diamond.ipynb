```python
nom_col=[1,2]
ord_col=[0]
sc_col=[6]
```

```python
trans_data=make_column_transformer((OneHotEncoder(sparse=False),nom_col),
                                   (OrdinalEncoder(),ord_col),
                                   (StandardScaler(),sc_col),remainder='passthrough')
```

```python
model=LinearRegression()
```

```python
set_config(display='diagram')
pipe=make_pipeline(trans_data,model)
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
pred=pipe.predict(x_test)
```

```python
mean_squared_error(pred,y_test)*100
```
