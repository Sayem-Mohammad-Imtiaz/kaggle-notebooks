```python
data = pd.read_csv('../input/novel-corona-virus-2019-dataset/covid_19_data.csv')
data.head(2)
```

```python
data[["Confirmed","Deaths","Recovered"]] =data[["Confirmed","Deaths","Recovered"]].astype(int)
```

```python
data['Active_case'] = data['Confirmed'] - data['Deaths'] - data['Recovered']
```

```python
selected_country = np.array(['Tunisia'])
selected_country
```

```python
from os import path
```

```python
i=1
for country in selected_country:
    data_con = data [(data['Country/Region'] == country) ].reset_index(drop=True)
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=data_con["ObservationDate"], y=data_con['Confirmed'],
                             mode="lines+text",name='Confirmed cases',marker_color='orange',))
    fig.add_annotation(
                x="03/22/2020", y=data_con['Confirmed'].max(),text="COVID-19 pandemic lockdown in {}".format(country),
                font=dict(size=15,color="red"))
    fig.add_shape( dict(type="line", x0="03/22/2020", y0=data_con['Confirmed'].max(), x1="03/22/2020",
                   line=dict(color="red",width=3) ))
    fig.add_annotation(x="04/22/2020", y=data_con['Confirmed'].max()-70, text="Month after lockdownin {}".format(country),
                      font=dict(family="Courier New, monospace",size=16,color="green"))
    fig.add_shape(dict( type="line", x0="04/22/2020", y0=data_con['Confirmed'].max(), x1="04/22/2020", 
                       line=dict(color="green",width=3)))
    fig.update_layout(title='Evolution of Confirmed cases over time in {}'.format(country), template='plotly_white')
    fig.write_image('confirmed_cases_in_{}.png'.format(country))
    fig.show()
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=data_con["ObservationDate"], y=data_con['Active_case'], mode="lines+text",
                             name='Active cases', marker_color='orange'))
    fig.add_annotation(x="03/22/2020",y=data_con['Active_case'].max(),text="COVID-19 pandemic lockdown in {}".format(country),
                       font=dict(family="Courier New, monospace",size=16,color="red"))
    fig.add_shape(dict(type="line",x0="03/22/2020",y0=data_con['Active_case'].max(),x1="03/22/2020",
                       line=dict(color="red",width=3)))
    fig.add_annotation(x="04/22/2020",y=data_con['Active_case'].max()-60,text="Month after lockdown in {}".format(country),
                       font=dict(family="Courier New, monospace",size=16,color="red"))
    fig.add_shape(dict(type="line",x0="04/22/2020",y0=data_con['Active_case'].max(),x1="04/22/2020",
                       line=dict(color="green",width=3)))
    fig.update_layout(title='Evolution of Active cases over time in {}'.format(country),template='plotly_white')
    fig.write_image('active_cases_in_{}.png'.format(country))
    fig.show()
    
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=data_con["ObservationDate"], y=data_con['Recovered'],mode="lines+text",
                             name='Recovered cases', marker_color='blue'))
    fig.add_annotation(x="03/22/2020",y=data_con['Recovered'].max(),text="COVID-19 pandemic lockdown in {}".format(country),
                            font=dict(family="Courier New, monospace",size=16,color="red"))
    fig.add_shape(dict(type="line",x0="03/22/2020",y0=data_con['Recovered'].max(),x1="03/22/2020",
                           line=dict(color="red",width=3)))
    fig.add_annotation(x="04/22/2020",y=data_con['Recovered'].max()-60,text="Month after lockdown in {}".format(country),
                           font=dict(family="Courier New, monospace",size=16,color="green"))
    fig.add_shape(dict(type="line",x0="04/22/2020",y0=data_con['Recovered'].max(),x1="04/22/2020",
                           line=dict(color="green",width=3)))
    fig.update_layout(title='Evolution of Recovered cases over time in Tunisia',template='plotly_white')
    fig.write_image('recovered_cases_in_{}.png'.format(country))
    fig.show()
```
