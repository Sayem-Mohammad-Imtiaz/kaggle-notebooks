**now let us  answer some quesionts **

```python
# what are the states having the highest and the lowest total cases
total_cases = df[['State/UTs','Total Cases']].sort_values(by=['Total Cases'],ascending=False).reset_index(drop=True)
plt.figure(figsize=(8,10))
ax =sns.barplot(x='Total Cases',y='State/UTs',data=total_cases)
ax.bar_label(ax.containers[0],padding=10,fmt='%g');
plt.show()
```

**from the figure above we can see that (Maharashtra) state has the highest total cases and (Andaman and Nicobar) state has the lowest number of cases**

```python
#what are the states having the highest number of deaths and lowest number of deaths?
deaths = df[['State/UTs','Deaths']].sort_values(by=['Deaths'],ascending=False).reset_index(drop=True)
plt.figure(figsize=(8,10))
ax =sns.barplot(x='Deaths',y='State/UTs',data=deaths)
ax.bar_label(ax.containers[0],padding=10,fmt='%g');
plt.show()
```

**** again (Maharashtra) state has the highest number of deaths and ( Dadra and Nagar Haveli and Daman and Diu)state has the lowest number of deaths****
