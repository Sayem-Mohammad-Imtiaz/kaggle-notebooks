```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv('/kaggle/input/ipl-data-set/matches.csv')
```

```python
# extract 1st five records
df.head()
```

```python
#extract last five records
df.tail()
```

```python
#check shape
df.shape
```

```python
# How many numbers of columns has in dataset
# Checking Null values
df.columns
```

```python
df.isnull().sum()
```

```python
#Remove unwanted columns
df.drop('umpire3',axis=1,inplace=True)
```

```python
df.head(1)
```

**So umpire3 has lots of null values and that was an unwanted column**

```python
# which city has how any matches?
plt.figure(figsize=(13,6))
df['city'].value_counts().plot(kind='bar')
```

```python
# Top 10 winning team
plt.figure(figsize=(13,6))
df['winner'].value_counts().plot(kind='barh')
```

```python
#Top 10 Player of the match player
df['player_of_match'].value_counts()[0:5].plot(kind='bar')
```

```python
# which team has won the toss many time?
plt.figure(figsize=(13,6))
df['toss_winner'].value_counts().plot(kind='bar')
```

```python
#how many numbers of time match winners is the toss winners?
plt.figure(figsize=(13,6))
df.loc[df.toss_winner==df.winner,'winner'].value_counts().plot(kind='barh')
```

```python
# how many numbers of time match winner is the toss losers?
df.loc[df.toss_winner!=df.winner,'winner'].value_counts().plot(kind='barh')
```
