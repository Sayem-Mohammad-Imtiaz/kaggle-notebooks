#choosing target


```python
y= heart_attack.output
```

#Choosing features

```python
heart_attack_features = ["age", "sex", "trtbps","fbs", "chol", "thalachh"]

X = heart_attack[heart_attack_features]
```

#Splitting data in to training and test

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,
                                                                random_state=0)





#train_X, val_X, train_y, val_y = train_test_split(X, y,random_state = 0)
```

#Random forest MAE


```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

heart_attack_model = RandomForestRegressor(random_state=1, n_estimators=1000)
heart_attack_model.fit(X_train, y_train)
heart_attack_pred1 = heart_attack_model.predict(X_valid)
print(mean_absolute_error(y_valid, heart_attack_pred1))
```

#RandomForestClassifier MAE

```python
from sklearn.ensemble import RandomForestClassifier

heart_attack_model3 = RandomForestClassifier(random_state=1, n_estimators=1000)
heart_attack_model3.fit(X_train, y_train)
heart_attack_pred3 = heart_attack_model3.predict(X_valid)
print(mean_absolute_error(y_valid, heart_attack_pred3))


```

#XGBRegresssor Mae

```python
from xgboost import XGBRegressor
heart_attack_model2 = XGBRegressor(n_estimators=500)
heart_attack_model2.fit(X_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(X_valid, y_valid)],
             verbose=False)
heart_attack_pred2 = heart_attack_model2.predict(X_valid)
mae_2 = mean_absolute_error(heart_attack_pred2, y_valid)
print("Mean Absolute Error:" , mae_2)
```
