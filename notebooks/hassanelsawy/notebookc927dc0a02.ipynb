```python
import pandas as pd
df = pd.read_csv('/kaggle/input/mobile-price-classification/train.csv')
df2 = pd.read_csv('/kaggle/input/mobile-price-classification/test.csv')
pd.read_csv('/kaggle/input/mobile-price-classification/train.csv')
#pd.read_csv('/kaggle/input/mobile-price-classification/test.csv')
```

```python
y = df["price_range"]
y.to_numpy
print(y)
```

```python
X = df.iloc[:,0:20]
X.to_numpy
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

parameters = {'n_neighbors':[2, 3, 4, 5, 6,], 'p':[1,2,3,4,5,10,20]}
knn = KNeighborsClassifier()
gsc = GridSearchCV(knn , parameters)
gsc.fit(X_train , y_train)
```

```python
gsc.best_params_
```

```python
knn = gsc.best_estimator_
y_pred = knn.predict(X_test)
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred , y_test)
```
