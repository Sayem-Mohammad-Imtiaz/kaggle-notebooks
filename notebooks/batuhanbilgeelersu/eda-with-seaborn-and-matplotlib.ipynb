# Exploratory Data Analysis with Seaborn and Matplotlib on Stroke Dataset

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from pandas.api.types import CategoricalDtype
```

```python
DataFrame = pd.read_csv("/kaggle/input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv")
df = DataFrame.copy()
```

```python
sns.boxplot(x = "bmi", data=df)
```

```python
df.bmi.fillna(df.bmi.median(),inplace=True)
```

```python
sns.distplot(df.bmi)
```

```python
smoking_categories = ["Unknown", "never smoked", "formerly smoked", "smokes"]
df.smoking_status = df.smoking_status.astype(CategoricalDtype(categories = smoking_categories, ordered = True))
```

```python
sns.barplot(x = "smoking_status", y = df.smoking_status.index, 
            palette = "Greys", data = df)
```

```python
sns.catplot(x = "stroke", y = "age", 
            palette = "cool", height=6.27, data=df)
```

```python
sns.catplot(x = "age",y = "smoking_status", kind = 'box', 
            hue = "gender", palette = "cool",height=6.27, data=df)
```

```python
sns.catplot(x="age", y="smoking_status", 
            hue="stroke",col="gender", aspect=.6,
            kind="box", palette = "cool", height=6.27, data=df)
```

```python

sns.catplot(x = "stroke", y = "bmi", 
            hue = "gender", kind = "swarm", 
            palette = "cool", 
            height=8.27, aspect=11.7/8.27, data=df);
```

```python
plt.figure(figsize=(15,15))
plt.subplot(2,1,1)
plt.title("Hypertension on BMI and Average Glucose Level")
sns.scatterplot(x = "avg_glucose_level", y = "bmi", 
                hue = "hypertension", palette = "OrRd", data=df)

plt.subplot(2,1,2)
plt.title("Hypertension on BMI and Age")
sns.scatterplot(x = "age", y = "bmi", 
                hue = "hypertension", palette = "PuBuGn", data=df)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,15))
plt.subplot(2,1,1)
plt.title("Stroke on BMI and Average Glucose Level")
sns.scatterplot(x = "avg_glucose_level" , y = "bmi", hue = "stroke", palette = "YlOrBr", data=df)

plt.subplot(2,1,2)
plt.title("Stroke on BMI and Age")
sns.scatterplot(x = "age", y = "bmi", 
                hue = "stroke", palette = "Reds", data=df)

plt.tight_layout()
plt.show()
```
