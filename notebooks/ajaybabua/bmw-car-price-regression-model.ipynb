# **Importing the Libaries**

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge

from sklearn.metrics import r2_score,mean_squared_error
```

```python
data = pd.read_csv("../input/used-car-dataset-ford-and-mercedes/bmw.csv")
data.head()
```

# **Encoding the data**

```python
encoder = LabelEncoder()
data['model'] = encoder.fit_transform(data['model'])
model_mapping = {index : label for index, label in enumerate(encoder.classes_)}
model_mapping
```

```python
data['transmission'] = encoder.fit_transform(data['transmission'])
transmission_mapping = {index : label for index, label in enumerate(encoder.classes_)}
transmission_mapping
```

```python
data['fuelType'] = encoder.fit_transform(data['fuelType'])
fuelType_mapping = {index : label for index, label in enumerate(encoder.classes_)}
fuelType_mapping
```

```python
x = data.drop('price', axis=1)
y = data['price']
```

# **Splitting the data**

```python
x_train,x_test, y_train, y_test =train_test_split(x, y, test_size=30, random_state=0)
```

# **LinearRegression**

```python
lin_model = LinearRegression()
lin_model.fit(x_train, y_train)

y_lin_model = lin_model.predict(x_test)
```

# **Lasso**

```python
l1_model = Lasso()
l1_model.fit(x_train,y_train)

y_l1_model = l1_model.predict(x_test)
```

# **Ridge**

```python
l2_model = Ridge()
l2_model.fit(x_train,y_train)

y_l2_model = l2_model.predict(x_test)
```

# **Finding Score for all Models**

```python
print(f'Linear score={lin_model.score(x_train, y_train)}')
print(f'Lasso score={l1_model.score(x_train, y_train)}')
print(f'Ridge score={l2_model.score(x_train, y_train)}')
```

# **Finding R^2 for all Models**

```python
print(f'R2_Linear = {r2_score(y_test,y_lin_model)}')
print(f'R2_Lasso = {r2_score(y_test,y_l1_model)}')
print(f'R2_Ridge = {r2_score(y_test,y_l2_model)}')
```

# **Finding RMS for all Models**

```python
print(f'RMSE_Linear = {np.sqrt(mean_squared_error(y_test,y_lin_model))}')
print(f'RMSE_Lasso = {np.sqrt(mean_squared_error(y_test,y_l1_model))}')
print(f'RMSE_Ridge = {np.sqrt(mean_squared_error(y_test,y_l2_model))}')
```
