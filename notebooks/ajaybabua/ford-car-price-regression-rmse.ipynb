# **Importing Labaries**

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso

from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv("../input/used-car-dataset-ford-and-mercedes/ford.csv")
data.head()
```

# **Encoding the data's**

```python
encoder = LabelEncoder()
data["model"] = encoder.fit_transform(data['model'])
model_mapping = {index : label for index, label in enumerate(encoder.classes_)}
model_mapping
```

```python
data['transmission'] = encoder.fit_transform(data['transmission'])
transmission_mapping = {index:label for index, label in enumerate(encoder.classes_)}
transmission_mapping
```

```python
data['fuelType'] = encoder.fit_transform(data['fuelType'])
fuelType_mapping = {index:label for index, label in enumerate(encoder.classes_)}
fuelType_mapping
```

```python
data
```

```python
X = data.drop("price", axis=1)
y = data["price"]
```

```python
X
```

# **Splitting data into Train and Test in the ratio of 70:30**

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.30, train_size=0.70, random_state=0)
```

```python
X_train
```

#  **Fitting train data's into Linear Regression, Lasso and Ridge**


```python
model_li = LinearRegression()
model_la = Lasso()
model_rg = Ridge()

model_li.fit(X_train, y_train)
model_la.fit(X_train, y_train)
model_rg.fit(X_train, y_train)
```

# **Finding Score for all Models**


```python
print(f'Linear score={model_li.score(X_train, y_train)}')
print(f'Lasso score={model_la.score(X_train, y_train)}')
print(f'Ridge score={model_rg.score(X_train, y_train)}')
```

# **Predicting the X_test data for all three Models**


```python
y_pred_li = model_li.predict(X_test)
y_pred_la = model_la.predict(X_test)
y_pred_rg = model_rg.predict(X_test)
```

# **Finding the Root Mean Square Error(RMSE) for all three Models**


```python
print(f'RMSE_Linear = {np.sqrt(mean_squared_error(y_test,y_pred_li))}')
print(f'RMSE_Lasso = {np.sqrt(mean_squared_error(y_test,y_pred_la))}')
print(f'RMSE_Ridge = {np.sqrt(mean_squared_error(y_test,y_pred_rg))}')
```

# **Finding the R^2 for all three Models**

```python
print(f'R2_Linear = {r2_score(y_test,y_pred_li)}')
print(f'R2_Lasso = {r2_score(y_test,y_pred_la)}')
print(f'R2_Ridge = {r2_score(y_test,y_pred_rg)}')
```
