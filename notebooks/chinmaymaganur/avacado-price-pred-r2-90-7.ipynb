**#lets check hoe avegare_price changes wrt to Seasons
spring runs from March 1 to May 31;
summer runs from June 1 to August 31;****
fall (autumn) runs from September 1 to November 30; and
winter runs from December 1 to February 28 (February 29 in a leap year).**

```python
conditions=[(df['month'].between(3,5,inclusive=True)),
           (df['month'].between(6,8,inclusive=True)),
           (df['month'].between(9,11,inclusive=True)),
           (df['month'].between(12,2,inclusive=True))]
values=['Spring','Summer','Fall','Winter']
df['seasons'] = np.select(conditions, values)
```

```python
#Season Price

aa=df.groupby('seasons')['AveragePrice'].mean().to_dict()
df_label['season_price']=df_label['seasons'].map(aa)
```

```python
#Season VOlume

bb=df_label.groupby('seasons')['Total Volume'].mean().to_dict()
df_label['season_Volume']=df_label['seasons'].map(bb)
df_label=pd.get_dummies(df_label,drop_first=True,columns=['type'])
```

```python
#region price  and avg yearly price

reg=df_label.groupby(['region'])['AveragePrice'].mean().to_dict()
df_label['region_price']=df_label['region'].map(reg)
reg=df_label.groupby(['year'])['AveragePrice'].mean().to_dict()
df_label['year_price']=df_label['year'].map(reg)
```

```python
col=['4046', '4225', '4770','Small Bags', 'Large Bags', 'XLarge Bags','Total Volume','Total Bags']
for i in col:
    df_label[i]=np.log(1+df_label[i])
```

```python
X=df_label[[ 'Total Volume','Total Bags', 'year','region', 'month', 'day', 'season_Volume',
       'type_organic', 'region_price', 'year_price']]
y=df_label['AveragePrice']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)


lr=LinearRegression()
lr.fit(X_train,y_train)
predict=lr.predict(X_test)
print('R2_score ',r2_score(y_test,predict))
print('mae ',mean_absolute_error(y_test,predict))
print('rmse ',np.sqrt(mean_squared_error(y_test,predict)))
sns.distplot((predict-y_test))
print(sum(predict-y_test))
```
