```python
df = timesData[timesData['year']==2011]
```

#  **LINE CHART (PLOT)** 


```python
import plotly.graph_objs as go

#Trace 1 
trace1 = go.Scatter(
                    x = df.world_rank,
                    y = df.citations,
                    mode= 'lines',
                    name = 'citations',
                    marker = dict(color = 'rgba(16,112,2,0.8)'),
                    text = df.university_name)

#Trace 2 
trace2 = go.Scatter(
                    x = df.world_rank,
                    y = df.teaching,
                    mode= 'lines+markers',
                    name = 'teaching',
                    marker = dict(color = 'rgba(80,20,80,0.8)'),
                    text = df.university_name)

data = [trace1,trace2]
layout = dict(title = 'Citiation and Teaching vs World Rank top 200',
             xaxis=dict(title='World Rank',ticklen=4,zeroline=False))
fig=dict(data = data, layout = layout)
iplot(fig)
```

# **SCATTER CHART** 

2014,2015 ve 2016 verilerinden ilk 100er tanesini alalım

```python
df2014 = timesData[timesData['year']==2014].iloc[:100,:]
df2015 = timesData[timesData['year']==2015].iloc[:100,:]
df2016 = timesData[timesData['year']==2016].iloc[:100,:]

#Trace 1
trace1 = go.Scatter(x = df2014.world_rank,
                   y=df2014.citations,
                   mode = "markers",
                   name = "2014",
                   marker = dict(color='rgba(255,0,0,0.8)'),
                   text=df2014.university_name)

#Trace 2
trace2 = go.Scatter(x = df2015.world_rank,
                   y=df2015.citations,
                   mode = "markers",
                   name = "2015",
                   marker = dict(color='rgba(0,255,0,0.8)'),
                   text=df2015.university_name)

#Trace 3
trace3 = go.Scatter(x = df2016.world_rank,
                   y=df2016.citations,
                   mode = "markers",
                   name = "2016",
                   marker = dict(color='rgba(0,0,255,0.8)'),
                   text=df2016.university_name)

data = [trace1,trace2,trace3]
layout= dict(title = 'Citation vs World Rank of Top 100 Universities with 2014,2015,2016',
            xaxis=dict(title = 'World Rank',ticklen = 4, zeroline=False),
            yaxis=dict(title = 'Citation',ticklen = 4, zeroline=False))
fig = dict(data = data, layout = layout)
iplot(fig)
```

# **BAR CHART**

```python
df2014 = timesData[timesData['year'] == 2014]. iloc[:3,:]

#Trace 1
trace1 = go.Bar(
                x = df2014.university_name,
                y = df2014.citations,
                name = "citations",
                marker = dict(color = 'rgba(255, 174, 255, 0.5)',
                             line=dict(color='rgb(0,0,0)',width=1.5)),
                text = df2014.country)
#Trace 2
trace2 = go.Bar(
                x = df2014.university_name,
                y = df2014.teaching,
                name = "teaching",
                marker = dict(color = 'rgba(255, 255, 128, 0.5)',
                              line=dict(color='rgb(0,0,0)',width=1.5)),
                text = df2014.country)
data = [trace1, trace2]
layout = go.Layout(barmode = "group")
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

Yan yana değil AltlıUstlu BAR CHART -dict ile yaptık bu sefer-

```python
x = df2014.university_name

trace1 = {
  'x': x,
  'y': df2014.citations,
  'name': 'citation',
  'type': 'bar'
};
trace2 = {
  'x': x,
  'y': df2014.teaching,
  'name': 'teaching',
  'type': 'bar'
};
data = [trace1, trace2];
layout = {
  'xaxis': {'title': 'Top 3 universities'},
  'barmode': 'relative',
  'title': 'citations and teaching of top 3 universities in 2014'
};
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

#  **PIE CHART** 

```python
df2016.head()
```

num_students degerleri float olmalıyken string olarak girilmis ve ondalık ayırmak icin nokta yerine virgül kullanilmis. 
Onlar duzeltiliyor.

```python
#data preprocessing
df2016 = timesData[timesData['year']==2016].iloc[:7,:]
pie1=df2016.num_students
pie1_list = [float(each.replace(',','.'))for each in df2016.num_students] #from str(2,4) to str(2.4) to float(2.4) 
labels = df2016.university_name
```

```python
fig = { "data" : [{
    "values" : pie1_list,
    "labels" : labels,
    "domain" : {"x" : [0, .5]},
    "name" : "Numberof Students Rate",
    "hoverinfo" : "label+percent+name",
    "hole" : .2,
    "type" : "pie"
}
],
      "layout" : {
          "title" : "Universities Number of Students Rates",
          "annotations" : [{
              "font" : {"size" : 20},
              "showarrow" : False,
              "text" : "Number Of Students",
              "x" : 0.20,
              "y" : 1
          }]
      }}

iplot(fig)
```

# **BUBBLE CHART**

```python
df2016 = timesData[timesData.year == 2016].iloc[:20,:]
num_students_size  = [float(each.replace(',', '.')) for each in df2016.num_students]
international_color = [float(each) for each in df2016.international]
data = [
    {
        'y': df2016.teaching,
        'x': df2016.world_rank,
        'mode': 'markers',
        'marker': {
            'color': international_color,
            'size': num_students_size,
            'showscale': True
        },
        "text" :  df2016.university_name    
    }
]
iplot(data)
```

# **HISTOGRAM**

```python
# preprocessing
x2011 = timesData.student_staff_ratio[timesData.year == 2011]
x2012 = timesData.student_staff_ratio[timesData.year == 2012]

#Visualization
trace1 = go.Histogram(
    x=x2011,
    opacity=0.75,
    name = "2011",
    marker=dict(color='rgba(171, 50, 96, 0.6)'))
trace2 = go.Histogram(
    x=x2012,
    opacity=0.75,
    name = "2012",
    marker=dict(color='rgba(12, 50, 196, 0.6)'))

data = [trace1, trace2]
layout = go.Layout(barmode='overlay',
                   title=' students-staff ratio in 2011 and 2012',
                   xaxis=dict(title='students-staff ratio'),
                   yaxis=dict( title='Count'),
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

# **SCATTER PLOT MATRİX**

```python
#import ff
import plotly.figure_factory as ff
#prepare data
dataframe = timesData[timesData.year==2015]
data2015 = timesData.loc[:,['research','international','total_score']]
data2015['index'] = np.arange(1,len(data2015)+1)
#Scatter Matrix
fig = ff.create_scatterplotmatrix(data2015,diag='box',index='index',colormap='Portland',
                                 colormap_type='cat',
                                 height=700,width=700)
iplot(fig)
```

# **INSET PLOT**

```python
# Trace 1
trace1 = go.Scatter(
    x=dataframe.world_rank,
    y=dataframe.teaching,
    name = "teaching",
    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
)
# Trace 2
trace2 = go.Scatter(
    x=dataframe.world_rank,
    y=dataframe.income,
    xaxis='x2',
    yaxis='y2',
    name = "income",
    marker = dict(color = 'rgba(160, 112, 20, 0.8)'),
)

data = [trace1,trace2]
layout = go.Layout(xaxis2=dict(domain = [0.6,0.95],
                               anchor = 'x2'),
                  yaxis2=dict(domain = [0.6,0.95],
                             anchor = 'y2'),
                    title = 'INCOME AND TEACHING VS WORLD RANK OF UNIVERSITTIES')

fig = go.Figure(data=data,layout=layout)
iplot(fig)
```

# **3D SCATTER PLOT**

```python
#Trace 1
trace1 = go.Scatter3d(x=dataframe.world_rank,
                     y=dataframe.research,
                     z=dataframe.citations,
                     mode='markers',
                     marker=dict(size=10,
                                color='rgb(255,0,0)'))

data = [trace1]
layout=go.Layout(margin=dict(l=0,
                             r=0,
                             b=0,
                             t=0))

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

# **MULTİPLE SUBPLOTS**

```python
#Trace 1
trace1 = go.Scatter(x=dataframe.world_rank,
                   y=dataframe.research,
                   name='research',
                   )
#Trace 2
trace2 = go.Scatter(x=dataframe.world_rank,
                   y=dataframe.citations,
                   name='citations',
                   xaxis='x2',
                   yaxis='y2'
                   )
#Trace 3
trace3 = go.Scatter(x=dataframe.world_rank,
                   y=dataframe.income,
                   name='income',
                   xaxis='x3',
                   yaxis='y3'
                   )
#Trace 4
trace4 = go.Scatter(x=dataframe.world_rank,
                   y=dataframe.total_score,
                   name='total_score',
                   xaxis='x4',
                   yaxis='y4'
                   )
data=[trace1,trace2,trace3,trace4]

layout = go.Layout(xaxis=dict(domain =[0,0.45]),
                   xaxis2=dict(domain=[0.55,1],anchor='y2'),
                   xaxis3=dict(domain=[0,0.45], anchor='y3'),
                   xaxis4=dict(domain=[0.55,1],anchor='y4'),
                   yaxis=dict(domain =[0,0.45]),
                   yaxis2=dict(domain=[0,0.45],anchor='x2'),
                   yaxis3=dict(domain=[0.55,1],anchor='x3'),
                   yaxis4=dict(domain=[0.55,1],anchor='x4'),
                   title='Research, Citation, Income and Total Score VS World Rank of Universities')

fig = go.Figure(data = data, layout = layout)
iplot(fig)
```
