In this project i will analyze some 911 call data:

The data contains these fields:
    
    lat:Latitude 
    lng:Longitude
    desc:Description of the emergency call
    zip:Zipcode
    title:Title
    timestamp:YYY-MM-DD HH:MM:SS
    twp:Township
    addr:Address

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
%matplotlib inline
```

```python
df=pd.read_csv('../input/montcoalert/911.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['zip'].value_counts().head()
```

```python
df['twp'].value_counts().head()
```

```python
df['title'].nunique()
```

```python
title=df['title'].iloc[0]
```

```python
title.split(':')[0]
```

```python
df['reason']=df['title'].apply(lambda title:title.split(':')[0])
```

```python
df['reason'].value_counts()
```

```python
sb.countplot(data=df,x='reason')
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp']=pd.to_datetime(df['timeStamp'])
```

```python
time=df['timeStamp'].iloc[0]
```

```python
time
```

```python
df['hour']=df['timeStamp'].apply(lambda time:time.hour)
```

```python
df['month']=df['timeStamp'].apply(lambda time:time.month)
```

```python
df['day of week']=df['timeStamp'].apply(lambda time:time.dayofweek)
```

```python
df.head()
```

```python
d={0:'Mon',1:'Thu',2:'Wed',3:'Thr',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['Day of week']=df['day of week'].map(d)
```

```python
df.drop('Day of week',axis=1,inplace=True)
```

```python
sb.countplot(x='day of week',data=df,hue='reason')
```

```python
sb.countplot(x='month',data=df,hue='reason')
```

```python
bymonth=df.groupby('month').count()
bymonth.head()
```

```python
bymonth['lat'].plot()
```

```python
sb.lmplot(x='month',y='twp',data=bymonth.reset_index())
```

```python
t=df['timeStamp'].iloc[0]
```

```python
t.date()
```

```python
df['date']=df['timeStamp'].apply(lambda t:t.date())
```

```python
df
```

```python
df.groupby('date').count()
```

```python
df.groupby('date').count()['lat'].plot()
plt.tight_layout()
```

```python
df[df['reason']=='Traffic'].groupby('date').count()['lat'].plot()
plt.tight_layout()
plt.title('Traffic')
```

```python
df[df['reason']=='Fire'].groupby('date').count()['lat'].plot()
plt.tight_layout()
plt.title('Fire')
```

```python
df[df['reason']=='EMS'].groupby('date').count()['lat'].plot()
plt.tight_layout()
plt.title('EMS')
```

```python
dayhour=df.groupby(['day of week','hour']).count()['reason'].unstack()
```

```python
plt.figure(figsize=(8,8),dpi=75)
sb.heatmap(dayhour,cmap='viridis')
```

```python
sb.clustermap(dayhour,cmap='viridis')
```

```python
daymonth=df.groupby(['day of week','month']).count()['reason'].unstack()
```

```python
plt.figure(figsize=(8,8),dpi=75)
sb.heatmap(daymonth,cmap='viridis')
```
