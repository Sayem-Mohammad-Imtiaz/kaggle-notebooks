### RANGE

```python
for i in range(10):
    print('Product: {} - Sale: ${}'.format(mobiles[i], sales[i]) )
```

```python
for i in range(5, 10):
    print('Product: {} - Sale: ${}'.format(mobiles[i], sales[i]) )
```

```python
i=0

cod_prod_price_list = []

for entry in enumerate(mobiles):
    temp = list(entry)
    temp.insert(3, sales[i])
    entry = tuple(temp)
    i+=1    
    cod_prod_price_list.append(entry)
```

```python
print(cod_prod_price_list)
```

```python
income = 0
for item in cod_prod_price_list:
    print(item)
```

```python
income = 0
for i, brand, sale in cod_prod_price_list:
    income += sale 
    
# juggle to format the thousands separator from dot (.) to comma (,), 
# and decimal separator from comma (,) to dot (.)
# What crazy engineering, huh?
income = "R${:_.2f}".format(income).replace('.', ',').replace('_', '.')

print("The total Revenue generated by the sale of goods, in Brazilian Real, was {}".format(income))
```
