```python
customer_data = pd.read_csv('/kaggle/input/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv')
customer_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = pd.get_dummies(customer_data.drop(['customerID','Churn'], axis=1))
y = customer_data['Churn'].apply(lambda x: 1 if x=='Yes' else 0)
```

```python
X_train, X_text, y_train, y_test = train_test_split(X, y, test_size=.2)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
# Seq - main model, load_model is so we can use the model from memory
from tensorflow.keras.models import Sequential, load_model
# Dense is how we'll add our layers to the model
from tensorflow.keras.layers import Dense
# How we are going to rate performance in this model
from sklearn.metrics import accuracy_score
```

```python
# First establish the model
model = Sequential()

#Add some hidden layers to the model
model.add(Dense(units=32, activation='relu', input_dim=len(X_train.columns)))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
# compile the model
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

```python
model.fit(X_train, y_train, epochs=200, batch_size=32)
```

```python
y_hat = model.predict(X_text)
y_hat = [0 if val <0.5 else 1 for val in y_hat]
```

```python
#y_hat
```

```python
accuracy_score(y_test, y_hat)
```

```python
model.save('tfmodels')
```
