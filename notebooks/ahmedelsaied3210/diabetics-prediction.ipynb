```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split,KFold,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
```

```python
dataset=pd.read_csv('../input/diabetes-dataset/diabetes2.csv')
dataset.head()
```

```python
X=dataset[dataset.columns]
X=X.drop(columns=['Outcome'])
X.head()
```

```python
Y=dataset['Outcome']
Y.head()
```

```python
from sklearn.feature_selection import SelectKBest,chi2
test=SelectKBest(score_func=chi2,k=4)
fit=test.fit(X,Y)
print(fit.scores_)
```

```python
X=X.drop(columns=['Pregnancies','BloodPressure','SkinThickness','DiabetesPedigreeFunction'])
```

```python
pd.DataFrame(X).head()
```

```python
from sklearn.preprocessing import StandardScaler
stand_x=StandardScaler()
X=stand_x.fit_transform(X)
```

```python
#histograms the dataset
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (15,20))
ax = fig.gca()
pd.DataFrame(X).hist(ax = ax)
```

```python
models=[]
models.append(('LR',LogisticRegression()))
models.append(('DT',DecisionTreeClassifier()))
models.append(('KN',KNeighborsClassifier()))
models.append(('NB',GaussianNB()))
models.append(('SVC',SVC()))
```

```python
results=[]
names=[]
scoring='accuracy'
for name,model in models:
    kfold=KFold(n_splits=10,random_state=42)
    cv_result=cross_val_score(model,X,Y,cv=kfold,scoring=scoring)
    results.append(cv_result)
    names.append(name)
    msg=("%s: %f (%f)" % (name,cv_result.mean(),cv_result.std()))
    print(msg)
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
fig.suptitle('Algorithms Coparison')
ax=fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
my_model=LogisticRegression()
my_model.fit(x_train,y_train)
```

```python
result=my_model.score(x_test,y_test)
print('Accuracy : ' ,(result*100))
```
