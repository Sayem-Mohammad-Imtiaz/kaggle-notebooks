```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#카테고리컬 변수들이 있으므로 label encoder import
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

#사용할 모델들
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.neural_network import MLPClassifier as MLP
```

```python
data = pd.read_csv('../input/diamonds/diamonds.csv', index_col='Unnamed: 0')
data.head()
```

```python
y=data['price']
X = data.drop(columns=['price'], axis=1)
```

```python
data
```

```python
#unique 함수도 잘 알아둘것!
len(data['cut'].unique())
```

```python
# f-string을 쓸때는 항상 큰 따옴표를 사용하자.
print(f"Total unique cuts are {len(data['cut'].unique())}")
print(f"Total unique color are {len(data['color'].unique())}")
print(f"Total unique clarity are {len(data['clarity'].unique())}")
```

```python
encoder = LabelEncoder()
X['cut'] = encoder.fit_transform(X['cut'])
X['color'] = encoder.fit_transform(X['color'])
X['clarity'] = encoder.fit_transform(X['clarity'])
```

```python
data
```

```python
# Attributes 와 Methods의 차이점. attributes는 인자를 받지 않고 속성을 보여주는 기능이고, method는 함수와 같은 기능 
encoder.classes_
```

```python
print({f"index : {i}":v for i, v in enumerate(encoder.classes_)})
```

```python
"""
categorical 한 변수들을 label encoder를 통해서 바꾸었는데 이를 minmaxscaler로 다시 조절할때는 조심해야 한다.
"""
scaler = MinMaxScaler()
X=scaler.fit_transform(X)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

```python
std_model=LinearRegression()
l1_model=Ridge(alpha=1)
l2_model=Lasso(alpha=1)
#nn_model = MLP(hidden_layer_sizes=(16,16), activation = 'relu', solver='adam', verbose=1)

std_model.fit(X_train, y_train)
l1_model.fit(X_train, y_train)
l2_model.fit(X_train, y_train)
#nn_model.fit(X_train, y_train)
```

```python
print(f"Without regularization : {std_model.score(X_test, y_test)}")
print(f"With ridge regularization : {l1_model.score(X_test, y_test)}")
print(f"With lasso regularization : {l2_model.score(X_test, y_test)}")
#print(f"With nn with batch size (32, 32): {nn_model.score(X_test)}")
```
