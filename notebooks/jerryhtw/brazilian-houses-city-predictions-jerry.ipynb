```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("../input/brasilian-houses-to-rent/houses_to_rent.csv")
data.head()
```

```python
data.shape
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
data
```

```python
data['floor'].replace(to_replace='-', value=0, inplace=True)
data
```

```python
data['animal'].replace(to_replace='acept', value=0, inplace=True)
data['animal'].replace(to_replace='not acept', value=1, inplace=True)

data['furniture'].replace(to_replace='furnished', value=0, inplace=True)
data['furniture'].replace(to_replace='not furnished', value=1, inplace=True)
```

```python
data
```

```python
for col in ['hoa','rent amount','property tax','fire insurance','total']:
    data[col].replace(to_replace='R\$', value='', regex=True, inplace=True)
    data[col].replace(to_replace=',', value='', regex=True, inplace=True)
```

```python
data.isin(['Sem info']).any()
```

```python
data['hoa'].replace(to_replace='Sem info', value=0, inplace=True)
data.isin(['Sem info']).any()
```

```python
data.isin(['Incluso']).any()
```

```python
data['hoa'].replace(to_replace='Incluso', value=0, inplace=True)
data['property tax'].replace(to_replace='Incluso', value=0, inplace=True)
data.astype(dtype=np.int64)
```

DataFrame의 모든 값들이 정상적으로 casting 되었음을 알 수 있다.

```python
data=data.sample(frac=1).reset_index(drop=True)
```

```python
data
```

```python
y=data['city']
X=data.drop(['city'], axis=1)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
scaler.fit(X)
X= scaler.transform(X)
```

```python
X
```

```python
X=pd.DataFrame(X)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, )
```

```python
y_train.shape
```

Try to use 3 models for the prediction today!!
1. Logistic regression, 2. SVM, 3. neural network

```python
from sklearn.linear_model import LogisticRegression as LR
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier as MLP
```

```python
log_model = LR(verbose=1)
svm_model = SVC(verbose=1)
nn_model = MLP(hidden_layer_sizes=(16,16), activation = 'relu', solver='adam', verbose=1)
```

```python
log_model.fit(X_train, y_train)
svm_model.fit(X_train, y_train)
nn_model.fit(X_train, y_train)
```

```python
print(log_model.score(X_test, y_test))
```

```python
print(svm_model.score(X_test, y_test))
```

```python
print(nn_model.score(X_test, y_test))
```

```python
#하지만!! // skewed data
#단순 score로는 별로 크게 말해주는 것이 없다. 데이터의 불균형이 심하기 때문
print(data[data.columns[0]].sum()/data.shape[0])
```

```python
##따라서 precision과 recall data의 조화 평균인 f1_score를 사용하자.
from sklearn.metrics import f1_score
```

```python
log_pred = log_model.predict(X_test)
svm_pred = svm_model.predict(X_test)
nn_pred = nn_model.predict(X_test)
```

```python
log_pred
```

```python
log_pred.shape
```

```python
print(f1_score(log_pred, y_test))
print(f1_score(svm_pred, y_test))
print(f1_score(nn_pred, y_test))
```
