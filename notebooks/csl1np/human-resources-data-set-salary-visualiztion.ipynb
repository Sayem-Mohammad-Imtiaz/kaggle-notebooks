Get data

```python
data = pd.read_csv("../input/human-resources-data-set/HRDataset_v14.csv")
data.info()
data.describe()
```

Salary overview

```python
data.Salary.describe() 
# 311 staff Max:250,000; Min:45,046 ; median:62,810
```

```python
sns.distplot(data.Salary)
plt.show()
```

Gender with salary

```python
sns.violinplot(x=data.Salary,y=data.Sex)
# Female overview have higher salary 
```

```python
Male = data[data["Sex"]=="M "]
Female= data[data["Sex"]=="F"]
sns.kdeplot(Male["Salary"],shade=True,label="M")
sns.kdeplot(Female["Salary"],shade=True,label="F")
```

Gender numeric view

```python
numeric = pd.DataFrame({
"Male": Male.describe()["Salary"],
"Female":Female.describe()["Salary"] })
numeric
# In numeric view average male have higher salary
```

In marriage status

```python
sns.boxplot(y=data.Salary,x=data.MaritalDesc,hue=data.MaritalDesc)
# Marriage status in single and married have higher salary than others
```

Marriage status numeric view

```python
Marriage_numeric = data[["MaritalDesc","Salary"]].groupby("MaritalDesc").describe()
Marriage_numeric
# From graph have shown different but the numeric view in 50% is not huge different, could be there data numbers are different
```

Department in salary

```python
sns.jointplot(x=data.Salary,y=data.Department)
# IT/IS have higher salary in overall
```

Department in numeric view

```python
DP = data[["Salary","Department"]].groupby("Department").describe()
DP
# Executive office could be ignore as only one person.
# In Median IT/IS and Software department have higher salary
```

Race with salary relationship

```python
sns.swarmplot(x=data.RaceDesc,y=data.Salary)
plt.xticks(rotation=90)
# can not really show the different as more staff is white people
```

Race in numeric view

```python
kk = data[["Salary","RaceDesc"]].groupby("RaceDesc").describe()
tt= data[["Salary","RaceDesc"]].describe()
kk
# In this data salary median is 62,810, compare in different race in median is very close, Hispanic only one person could ignore
```
