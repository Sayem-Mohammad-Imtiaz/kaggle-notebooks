Fill in with mean values in missing places

```python
df.fillna(df.mean(), inplace=True)
```

```python
#Target variable : 'TenYearCHD'

print(df.groupby('TenYearCHD').size())
```

Split the data for target variable and dependant variables

```python
X = df.drop(columns=['TenYearCHD'])
```

```python
X.head()
```

```python
y = df['TenYearCHD']
```

```python
y.head()
```

import train_test_split and Logistic Regression

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

Fit the model

```python
lr.fit(X_train, y_train)
```

Predict the model

```python
predictions = lr.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
```

**Obtained Accuracy of 86%**

```python
print(classification_report(y_test, predictions))
```
