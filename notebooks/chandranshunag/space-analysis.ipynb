In this notebook we are going to collect data about various space missions from 1957. For that we will be using a kaggle dataset called "all space missions from 1957"

## Load data
first we have to load the data present in the dataset it is available at ../input/all-space-missions-from-1957. We will be using a file in it called Space_Corrected.csv at ../input/all-space-missions-from-1957/Space_Corrected.csv

```python
import pandas as pd

# The location of the file
fileloc = '../input/all-space-missions-from-1957/Space_Corrected.csv'

# convert the data into a pandas dataframe
data = pd.read_csv(fileloc)
```

now show the data

```python
companies = data["Company Name"]
comp = companies.tolist()
l = []
for i in comp:
    if not i in l:
        l.append(i)
l
```

```python
dic = {}
for i in l:
    missions = data[data["Company Name"] == i]
    dic[i] = missions.shape[0]
dic
```

OK, we should have ploted it but just trying to plot them the hundreds of labels are making it very hard to see.
We should but instead we would just show the 9 most launched companies and the rest will be shown in a whole other group named "other".

```python
dic = sorted(dic.items(), key = lambda kv:(kv[1], kv[0]))
dic
```

As you see we have sorted it in ascending order least missions to greatest missions

```python
dec = list(reversed(list(dic)))
dec
```

We are succesful in showing the launches in decending order now we will plot it in our graph

```python
counter = 1
bigs = {}
other = 0
for i in dec:
    j = i[1]
    i = i[0]
    if not counter > 9:
        bigs[i] = j
    else:
        other = other + j
    counter = counter + 1
```

Extract the names and the values and save them in two lists
