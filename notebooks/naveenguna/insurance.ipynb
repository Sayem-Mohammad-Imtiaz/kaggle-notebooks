```python
import scipy.stats as stat
import pylab
def check_dist(df, feature):
    plt.figure(figsize = (10,6))
    plt.subplot(1,2,1)
    sns.distplot(df[feature])
    plt.subplot(1,2,2)
    stat.probplot(df[feature], dist = 'norm', plot = pylab) 
    plt.show()
    
    
```

```python
print('Age:')
print(check_dist(df2,'age'))
print('BMI:')
print(check_dist(df2,'bmi'))
print('Charges:')
print(check_dist(df2,'charges'))
```
