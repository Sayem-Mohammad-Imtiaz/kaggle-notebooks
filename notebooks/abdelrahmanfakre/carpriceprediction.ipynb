```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('/kaggle/input/vehicle-dataset-from-cardekho/Car details v3.csv')
data.head()
```

```python
data.isnull().sum(),data.shape
```

```python
data.dropna(axis = 0,inplace = True)
```

```python
data.drop('owner',axis = 1,inplace = True)
data.drop('mileage',axis = 1,inplace = True)
data.drop('seller_type',axis = 1,inplace = True)
data.drop('torque',axis = 1,inplace = True)
```

```python
data.dtypes
```

```python
data['name'] = data['name'].factorize()[0]
data['fuel'] = data['fuel'].factorize()[0]
data['transmission'] = data['transmission'].factorize()[0]
```

```python
number = []
for word in data['engine']:
    number.append((float)(word.split()[0])) 
data['engine'] = number
```

```python
data['engine']
```

```python
number = []
for word in data['max_power']:
    number.append((float)(word.split()[0])) 
data['max_power'] = number
```

```python
data['max_power']
```

```python
data.dtypes
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain , Xtest , ytrain, ytest = train_test_split(data.drop('selling_price',axis = 1),data['selling_price'],test_size = 0.2,random_state = 41)
```

# Normalization

```python
from sklearn.preprocessing import MinMaxScaler
scaller = MinMaxScaler()
Xtrain = scaller.fit_transform(Xtrain)
Xtest = scaller.fit_transform(Xtest)
```

# Model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(Xtrain,ytrain)
ypred = model.predict(Xtest)
compare = pd.DataFrame()
compare['ypred'] = ypred
compare['ytest'] = ytest.reset_index()['selling_price']
compare
```

```python
from sklearn.metrics import mean_squared_error

RMS = mean_squared_error(ytest,ypred,squared = False)
R1 = model.score(Xtrain,ytrain)
R2 = model.score(Xtest,ytest)
RMS,R1,R2
```
