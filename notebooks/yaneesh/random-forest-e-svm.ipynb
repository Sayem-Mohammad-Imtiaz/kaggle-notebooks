## 1. Importando os dados

```python
import pandas as pd
```

```python
df = pd.read_csv('../input/breast-cancer-wisconsin-data/data.csv')
```

```python
df.head()
```

## 2. Pré-processamento 

```python
df.isnull().sum()
```

```python
df_1 = df.drop(columns = "Unnamed: 32")
df_1.head()
```

### Alterando Benígno e Malígno para 0 e 1

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df_1['diagnosis'] = le.fit_transform(df_1['diagnosis'])
```

```python
df_1.head()
```

```python
exams = df_1.drop(columns = ['id', 'diagnosis'])
diagnosis = df_1.diagnosis
```

```python
from sklearn.model_selection import train_test_split
from numpy import random

SEED = 10
random.seed(SEED)
treino_x, teste_x, treino_y, teste_y = train_test_split(exams, diagnosis)
```

## 3. Machine Learning - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(treino_x, treino_y)
acc = clf.score(teste_x, teste_y)*100
print(" %.2f%%" % acc)
```

## 4. Machine Learning - SVM

```python
from sklearn import svm
clf_1 = svm.SVC()
clf_1.fit(treino_x, treino_y)
acc_1 = clf_1.score(teste_x, teste_y)*100
print(" %.2f%%" % acc_1)
```
