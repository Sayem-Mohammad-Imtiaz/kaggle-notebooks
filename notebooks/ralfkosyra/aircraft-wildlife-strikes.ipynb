```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/wildlife-strikes/database.csv')
```

```python
strike_damage = {}
for column in df.columns:
    col_sep = column.split(' ')
    #print(len(col_sep), col_sep)
    if len(col_sep) > 1 and col_sep[1] == 'Strike':
        strike_damage[col_sep[0]] = df[col_sep[0] + ' Damage'].sum() / df[column].sum()                        
```

```python
#Airplane part probability to be damaged
plt.bar(list(strike_damage.keys()),list(strike_damage.values()))
plt.xticks(rotation=90)
print(max(strike_damage,key=strike_damage.get))
```

```python
#Bird species probability to strike
#print(df['Species ID'].value_counts())
#x=df['Species ID'].unique()
birds=dict(df['Species ID'].value_counts(ascending=False))

x = list(birds.keys())[:15]
y = list(birds.values())[:15]

print(x[y.index(max(birds.values()))],max(birds.values()))
plt.barh(x,y)
```

```python
any_damage = []

for index, row in df.iterrows():
    #print(row['Nose Damage'])
    sum = 0
    for column in df.columns:
        col_sep = column.split(' ')
        if len(col_sep) > 1 and col_sep[1] == 'Damage':
            #print(row[column])
            if row[column] > 0:
                sum += row[column]
                break
    any_damage.append(sum) 
any_damage = np.array(any_damage)
```

```python
df['Any Dmg'] = any_damage
```

```python
df_dmg = df[df['Any Dmg']==1]
```

```python
#Bird species probability to cause damage
birds_dmg=dict(df_dmg['Species ID'].value_counts(ascending=False))

x2 = list(birds_dmg.keys())[:15]
y2 = list(birds_dmg.values())[:15]

print(x2[y2.index(max(birds_dmg.values()))],max(birds_dmg.values()))
plt.barh(x2,y2)
```
