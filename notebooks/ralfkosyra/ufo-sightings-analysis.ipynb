```python
import datetime
import math
```

```python
for i, dt in enumerate(df['datetime']):
    try:
        dt = datetime.datetime.strptime(dt, '%m/%d/%Y %H:%M')
        #print(dt)
    except:
        #print(dt)
        dt = dt.replace('24','23')
        dt = datetime.datetime.strptime(dt, '%m/%d/%Y %H:%M')
        dt += datetime.timedelta(hours=1)
        #print(dt)
        #break
    df['datetime'][i] = dt
    #print(dt)
    #break
```

```python
print(df['datetime'][0])
```

```python
month_counts = [0] * 12
season_counts = {'Winter':0,'Spring':0,'Summer':0,'Autumn':0}
hour_counts = [0] * 24
daytime_counts = {'Night':0,'Dawn':0,'Day':0,'Dusk':0}
for dt in df['datetime']:
    month = dt.month
    month_counts[month-1] += 1 
    if month in (1,2,3):
        season_counts['Winter'] += 1
    elif month in (4,5,6):
        season_counts['Spring'] += 1
    elif month in (7,8,9):
        season_counts['Summer'] += 1
    elif month in (10,11,12):
        season_counts['Autumn'] += 1
        
    hour = dt.hour
    hour_counts[hour] +=1
    if hour in (21,22,23,0,1,2):
        daytime_counts['Night'] += 1
    elif hour in range(3,9):
        daytime_counts['Dawn'] += 1
    elif hour in range(9,15):
        daytime_counts['Day'] += 1
    elif hour in range(15,21):
        daytime_counts['Dusk'] += 1
        
    
month_count_errs = [math.sqrt(mc) for mc in month_counts]
season_count_errs = {'Winter':0,'Spring':0,'Summer':0,'Autumn':0}
for season, sc in season_counts.items():
    season_count_errs[season] = math.sqrt(sc) 
hour_count_errs = [math.sqrt(hc) for hc in hour_counts]
daytime_count_errs = {'Night':0,'Dawn':0,'Day':0,'Dusk':0}
for daytime, dc in daytime_counts.items():
    daytime_count_errs[daytime] = math.sqrt(dc)
```

```python
plt.bar(range(1,len(month_counts)+1),month_counts,yerr=month_count_errs)
plt.show()
```

```python
plt.bar(season_counts.keys(),season_counts.values(),yerr=season_count_errs.values())
plt.show()
```

```python
plt.bar(range(len(hour_counts)),hour_counts,yerr=hour_count_errs)
plt.show()
```

```python
plt.bar(daytime_counts.keys(),daytime_counts.values(),yerr=daytime_count_errs.values())
plt.show()
```

```python
lon = df['longitude ']
lat = df['latitude']
```

```python
for i, val in enumerate(lat):
    if type(val) is str:
        val = float(''.join(c for c in val if (c.isdigit() or c == '.')))
        lat[i] = val    
```

```python
# Define the borders
deltaX = (max(lon) - min(lon))/10
deltaY = (max(lat) - min(lat))/10
xmin = min(lon) - deltaX
xmax = max(lon) + deltaX
ymin = min(lat) - deltaY
ymax = max(lat) + deltaY

print(xmin, xmax, ymin, ymax)
# Create meshgrid
xx, yy = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
```

```python
from scipy.stats import kde
import matplotlib.colors as colors
```

```python
lon = lon.astype(np.float64)
lat = lat.astype(np.float64)
nbins=100
k = kde.gaussian_kde([lon,lat])
xi, yi = np.mgrid[lon.min():lon.max():nbins*1j, lat.min():lat.max():nbins*1j]
zi = k(np.vstack([xi.flatten(), yi.flatten()]))
```

```python
plt.pcolormesh(xi, yi, zi.reshape(xi.shape), shading='auto', cmap='RdBu_r',
               norm=colors.LogNorm(vmin=zi.min(), vmax=zi.max()))
```
