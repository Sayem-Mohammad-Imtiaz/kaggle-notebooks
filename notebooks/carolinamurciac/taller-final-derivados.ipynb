 **Número de canciones por artista**
>  Primero, se analizará cuantas caciones por artista hay en el conteo

```python
my_data[["Artist.Name","Track.Name"]].groupby("Artist.Name").count() 

```

```python
nca=my_data.groupby(["Artist.Name"]).count().iloc[:,:1]
nca.columns=['Number of Songs']
print(nca)
```

```python
import matplotlib.pyplot as plt
from itertools import cycle, islice
df = pd.DataFrame(nca)
fig, ax = plt.subplots()
fig.set_size_inches(6,6)
df.plot.barh(stacked=True, ax=ax, color='pink')
ax.set_title("Número de canciones por artista")
plt.xlabel("Nombre Artista")
plt.xlabel("Número de canciones")


```

Con la tabla y la gráfica podemos observar que la mayoria de artistas de la lista tienen una canción en esta. Existen ecepciones como J Blavin, Billie Eilish, Bad Bunny, entre otrso que tienen 2 caciones en la lista y un dato atipico como Ed Sheeran quin tiene 4 canciones en la playlist.

 **Mayor Popularidad**

 **Beats por minuto **
>  Cuarto, se estudia los beast por minuto de cada canción.Los beats por minuto  es una unidad empleada para medir el ritmo en música, equivale al número de pulsaciones que caben en un minuto.

```python
my_data[["Track.Name","Beats.Per.Minute"]]
```

```python
sns.distplot(my_data["Beats.Per.Minute"],  color="green")
plt.title("Beats por Canción")
```

Con el anterior gráfico podemos ver que hay una mayor frecuencia  alrededor de los 90-100 beats por minutos.Además, despues de este rango hay un registro de frecuencias que van disminuyendo a medida que aumenta los beats por minuto.

 **Energía por Canción vs si es Bailable**
>  Quinto, se intenta averiguar si  hay una relacion entre la energía de una canción y si esta es bailable. Ambas variables están medidas [1-100]

```python
my_data[["Track.Name","Energy","Danceability"]]
```

```python
plt.plot(my_data["Danceability"], color="r")  
plt.xlabel("Top de canciones")
plt.ylabel("Valor")
plt.ioff()   
plt.plot(my_data["Energy"], color="y") 
plt.ion()
plt.title("Energía por Canción vs si es Bailable")
plt.legend()
  
```

A partir del gráfico se puede evidenciar que puede haber una relación directamente proporcional entre la energía de una canción y si es bailable. No obstante, esta relación no es hace presente en todas las canciones de la lista, esto se puede deber a que dentro de la lista hay canciones que son menos enérgicas (canciones lentas) que serían por supuesto lo opuesto al tipo de cación bailable. 

 **¿Si la canción es bailable es más popular?**
>  Por último, se intenta averiguar si  hay una relacion entre la popularidad de una canción y si esta es bailable. ambas variables estan medidas [1-100]% 

```python
my_data[["Track.Name","Popularity","Danceability"]]
```

```python
plt.scatter("Danceability", "Popularity", data=my_data[["Track.Name","Popularity","Danceability"]], color="c")
plt.title("¿Si la canción es bailable es más popular?")
plt.xlabel("Bailable")
plt.ylabel("Popularidad")
plt.show()
```

Finalmente, se observa con el gráfico que si hay una relacion entre si una cación es bailble y su popularidad, se podria determinar que si una cación es bailable puede ser un determinante de su popularidad.No obstante, tambien se observa pocos casos en los que la cancion tiene un mayor nivel de popularidad pero es poco bailable o lo opuesto tiene un gran grado de bailabilidad pero es menos popular entre las canciones de la lista.
