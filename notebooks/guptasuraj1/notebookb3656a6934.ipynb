```python
import pandas as pd
path="../input/vehicle-dataset-from-cardekho/car data.csv"

df=pd.read_csv(path)
#print(df.head())
#print(df['Seller_Type'].unique())
#print(df['Transmission'].unique())
#print(len(df['Car_Name'].unique()))
#print(df['Owner'].unique())

#finding the misising values and cells
#df.isnull().sum()
#finding colums
#print(df.columns)

#selecting new features and draw a new final datasets
final_df=df[['Year', 'Selling_Price', 'Present_Price', 'Kms_Driven',
       'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']]
 
final_df['current_year']=2020
final_df['years_old']=final_df['current_year']-final_df['Year']



final_df=final_df.drop(['Year','current_year'],axis=1)
final_df=pd.get_dummies(final_df,drop_first='true')
final_df.head(6)
```

```python
Y=final_df['Selling_Price']                  #dependent features 
X=final_df.drop(['Selling_Price'],axis=1)  
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=51)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
lr=RandomForestRegressor()
lr.fit(X_train,Y_train)
lr.score(X_test,Y_test)
```
