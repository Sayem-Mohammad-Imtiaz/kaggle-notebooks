Suggestions and comments are welcome.

```python
# importing dependancies
import numpy as np
import pandas as pd
from sklearn import linear_model, preprocessing # for normalizing labels
from sklearn.svm import SVC # estimator
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# taking a look at the data
df = pd.read_csv("/kaggle/input/star-dataset/6 class csv.csv")
df
```

```python
print(df['Star color'].unique())
print(df['Spectral Class'].unique())
```

```python
# cleaning
df['Star color'] = df['Star color'].replace('Blue white', 'Blue White').replace('Blue-white', 'Blue White').replace('Blue white ', 'Blue White').replace('Blue-White', 'Blue White').replace('Blue ', 'Blue').replace('white', 'White').replace('Whitish', 'White').replace('Yellowish White', 'Yellow White').replace('yellow-white', 'Yellow White').replace('White-Yellow', 'Yellow White').replace('yellowish', 'Yellow').replace('Yellowish', 'Yellow').replace('Pale yellow orange', 'Yellow Orange').replace('Orange-Red', 'Orange Red')

# normalizing
le = preprocessing.LabelEncoder()
df['Spectral Class'] = le.fit_transform(list(df['Spectral Class']))
df['Star color'] = le.fit_transform(list(df['Star color']))
```

```python
# after normalizing
print(df['Star color'].unique())
print(df['Spectral Class'].unique())
```

```python
# using SVC with 'linear' kernel
X = df.drop(columns='Star type')
y = df['Star type']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
model = SVC(kernel="linear", C=0.1)
model.fit(X_train, y_train)
predictions = model.predict(X_test)
score = accuracy_score(y_test, predictions)
print('Accuracy:',score*100,'%')
```
