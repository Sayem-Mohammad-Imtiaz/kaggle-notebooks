```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

```python
df=pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
sns.pairplot(df,hue='Outcome')
```

```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
col=df.columns[0:8]
col
```

```python
scaler=StandardScaler()
x=pd.DataFrame(scaler.fit_transform(x),columns=col)
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
k=np.arange(2,30)
train_acc=np.empty(len(k))
test_acc=np.empty(len(k))

for i,k in enumerate(k):
    knn=KNeighborsClassifier(k)
    knn.fit(x_train,y_train)
    train_acc[i]=knn.score(x_train,y_train)
    test_acc[i]=knn.score(x_test,y_test)
```

```python
k=np.arange(2,30)
```

```python
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(k,train_acc,color='r',marker='o')
plt.title('k-value vs Train_acc')
plt.xlabel('K-value')
plt.ylabel('Train_acc')
plt.subplot(1,2,2)
plt.plot(k,test_acc,color='r',marker='o')
plt.title('k-value vs Test_acc')
plt.xlabel('K-value')
plt.ylabel('Test_acc')
```

```python
idx=np.where(test_acc==max(test_acc))
a=k[idx]
a
```

```python
model=KNeighborsClassifier(n_neighbors=21)
model.fit(x_train,y_train)
```

```python
ypred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,ypred)
```

```python
confusion_matrix(y_test,ypred)
```

```python
print(classification_report(y_test,ypred))
```
