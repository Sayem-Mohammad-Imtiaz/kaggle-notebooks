```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
vg=pd.read_csv("../input/vgsales.csv")
```

```python
vg
```

```python
vg.dtypes
```

```python
vga=vg.dropna()
```

```python
vga.Year = vga.Year.astype(int)
vga.dtypes
```

```python
sales=vga.groupby(["Year"])["NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"].sum().reset_index()
sales
```

```python
sales1=vga.groupby(["Year","Platform"])["Global_Sales"].sum().reset_index()
sales1.drop("Year", axis=1).boxplot(by="Platform", figsize=(12, 6))
plt.xticks(rotation=90)
```

```python
fig, ax = plt.subplots(figsize=(12, 5))
sns.barplot(x='Year',y='Global_Sales',data=sales)
ax.set(xlabel='Year on Year', ylabel='Global Sales')
plt.xticks(rotation=90)
```

```python
platf1=sales1[sales1.Platform=="PS"]
platf2=sales1[sales1.Platform=="PS2"]
platf3=sales1[sales1.Platform=="PS3"]
platf4=sales1[sales1.Platform=="PS4"]

platf=pd.concat([platf1,platf2,platf3,platf4])
platf
```

```python
sns.pairplot(sales1)
```

```python
platf.Year = platf.Year.astype(object)
```
