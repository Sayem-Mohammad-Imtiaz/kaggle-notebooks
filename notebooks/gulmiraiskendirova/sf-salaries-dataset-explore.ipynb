# Exploring SF Salaries Dataset. Finding unique insights.

```python
df = pd.read_csv('../input/sf-salaries/Salaries.csv', low_memory=False)
```

* Finding number of entries there are by using .info() method 

```python
df.info()
```

As it can be seen, BasePay, OvertimePay, OtherPay and Benefits are objects which is not right, so we have to convert it to numerical values. As this columns might also contain non numerical values, errors='coerce' attribute for coerce invalid values to NaN will be used.

```python
objlist = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits']

for obj in objlist:
    df[obj] = pd.to_numeric(df[obj], errors = 'coerce')
```
