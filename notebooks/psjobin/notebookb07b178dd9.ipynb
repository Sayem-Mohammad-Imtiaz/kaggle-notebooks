```python

class Animal:  # Animal class
   default_name = "AN ANIMAL"  # this is a class level attribute

   @staticmethod
   def roar(x, y):
       print("Roar "+x+y)

   @classmethod
   def adultAnimal(cls):
       return cls(21)


   def __init__(self, age):  # class constructor
       self.age = age  # this is an instance attribute


   def sayHi(self):  # this is an class method
       print(f"Hello, I'm an animal and I am {self.age} year(s) old.")
```

code above.

```python
# create a class of animal
animal = Animal(2)
animal.sayHi()
```
