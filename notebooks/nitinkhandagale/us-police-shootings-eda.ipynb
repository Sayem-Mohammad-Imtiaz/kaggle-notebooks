```python
import pandas as pd
import sidetable
```

```python
df = pd.read_csv('../input/us-police-shootings/shootings.csv')
```

```python
df.head()
```

```python
df.stb.missing()
```

```python
df = df.drop(['id','name'], axis='columns')
```

```python
df.stb.missing()
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
race_group = df.stb.freq(['race'])

fig = px.pie(race_group, names='race', values='count', color_discrete_sequence=px.colors.qualitative.Antique)
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title='Police Shootings by Race')
fig.show()
```

```python
df.stb.freq(['race','gender'])
```

```python
race_gender_group = df.stb.freq(['gender'])

fig = go.Figure(go.Bar(x=race_gender_group['gender'], y=race_gender_group['count'],
             marker_color =px.colors.qualitative.Prism))

fig.update_traces(textposition='outside', text=race_gender_group['count'])


fig.update_layout(title='Genders Distribution')
fig.show()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
state_data = df.stb.freq(['state'])

fig = go.Figure(go.Bar(
    x=state_data['state'],
    y = state_data['percent'],
    marker_color=px.colors.sequential.Plasma_r
))

fig.update_layout(title = 'Statewise Numbers', yaxis=dict(title='Shootings percent'))
fig.show()
```

```python
city_data = df.stb.freq(['city'], thresh=9)
city_data.drop(city_data.tail(1).index,inplace=True)

fig = go.Figure(go.Bar(
    x=city_data['city'],
    y = city_data['percent'],
    marker_color=px.colors.sequential.RdBu
))

fig.update_layout(title = 'Top most affected Cities', yaxis=dict(title='Shootings percent'))
fig.show()
```

```python
 illness_measure = df.stb.freq(['signs_of_mental_illness'])
 illness_measure
```

```python
fig = px.pie(illness_measure, names='signs_of_mental_illness', values='percent',
             color_discrete_sequence = px.colors.qualitative.Safe)

fig.update_traces(textposition='inside', textinfo='percent+label',pull=[0, 0.1])
fig.update_layout(title='Mental Illness Measure')
fig.show()
```

```python
bodycam = df.stb.freq(['body_camera'])
bodycam
```

```python
fig = px.pie(bodycam, names='body_camera', values='percent',
             color_discrete_sequence = px.colors.qualitative.Vivid)

fig.update_traces(textposition='inside', textinfo='percent+label', pull = [0,0.1])
fig.update_layout(title='Was Officer had Body Camera ? ')
fig.show()
```
