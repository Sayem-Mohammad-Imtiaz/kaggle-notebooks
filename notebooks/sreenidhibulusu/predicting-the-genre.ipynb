Fro here I started preparing data for the MLA, I used the MLA's to predict if a books is a FICTION or NON-FICTION, In the cell below I converted Non-fiction to 0, fiction to 1. 

```python
df['Genre'].replace(to_replace=['Non Fiction','Fiction'], value=[1,0],inplace=True)
df.head()
```

```python
X = df[['User Rating', 'Reviews', 'Price', 'Year']]
print(X[0:5])
y = df['Genre'].values
y[0:5]
```

Since the algorithms I'll use use mostly numerical data as weights, I seperated all the columns with numerical value colums to the parameters (X) and the 'Genre' col to be the predicted values (Y)

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

```python
#prediction from logistic regression

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.25, shuffle = False)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR
yhat = LR.predict(X_test)
print(yhat)

yhat_prob = LR.predict_proba(X_test)
```

```python
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
jaccard_score(y_test, yhat,pos_label=0)
```
