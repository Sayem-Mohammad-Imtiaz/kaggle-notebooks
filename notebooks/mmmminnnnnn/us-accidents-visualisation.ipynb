# Visualise Data
## import packages

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

use pie chart to check how the accidents distributed according to state

```python
plt.figure(figsize =(10,5))
us_accidents.groupby(['State'])['Severity'].count().head(10).sort_values(ascending=False).plot.pie()
#plt.pie(us_accidents['State'].value_counts(), labels=lbs, autopct="%1.1f%%")
```

User bar chart to visualise the amount of accidents according to its severity.

```python
plt.title('Severity')
plt.xlabel('Severity')
plt.ylabel('Amount')
plt.xticks(np.arange(len(us_accidents['Severity'].value_counts())), us_accidents['Severity'].value_counts().index)
plt.bar(np.arange(len(us_accidents['Severity'].value_counts())), us_accidents['Severity'].value_counts())
```

Use line chart to visualise the amount of car accidents by time of the day.

```python
us_accidents["Start_Time"] = pd.to_datetime(us_accidents["Start_Time"])
us_accidents["Day_Time"] = us_accidents["Start_Time"].dt.hour
sns.countplot(x='Day_Time', data=us_accidents)
```

```python
plt.title("Amount of car accidents with time of the day")
plt.xlabel("Time of the day")
plt.ylabel("Amout of car accidents")
Day_Time_Counts = us_accidents["Day_Time"].value_counts(sort=False)
sns.lineplot(data=Day_Time_Counts)
```
