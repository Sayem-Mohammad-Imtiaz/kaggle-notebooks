```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/heart-disease-prediction-using-logistic-regression/framingham.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['education'].fillna(df['education'].median(),inplace=True)
df['cigsPerDay'].fillna(df['cigsPerDay'].median(),inplace=True)
df['BPMeds'].fillna(df['BPMeds'].median(),inplace=True)
df['totChol'].fillna(df['totChol'].median(),inplace =True)
df['BMI'].fillna(df['BMI'].median(),inplace=True)
df['heartRate'].fillna(df['heartRate'].median(),inplace=True)
df['glucose'].fillna(df['glucose'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
plt.figure(figsize=(12,8),dpi=200)
sns.scatterplot(data = df , x = 'age' , y='heartRate' ,hue ='TenYearCHD',alpha=0.6)
```

```python
sns.boxplot(x='TenYearCHD',y='age',data = df)
```

```python
plt.figure(figsize=(12,8),dpi= 200)
sns.heatmap(df.corr(),annot=True,cmap='Blues')
```

```python
# Now as we see there is imbalance data issue we have here and we should fix it before working on this data 
sns.countplot(data = df , x = 'TenYearCHD')
```

```python
X = df.drop('TenYearCHD',axis=1)
```

```python
y = df['TenYearCHD']
```

```python
from imblearn import over_sampling , under_sampling
```

```python
#We will use over sampling because of the data around 4000 rows so we should use oversampling
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler(random_state=0)
```

```python
X_resambled , y_resampled = ros.fit_resample(X,y)
```

```python
X_resambled.shape
```

```python
y_resampled.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resambled, y_resampled, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaled_x_train = scaler.transform(X_train)
```

```python
scaled_x_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(scaled_x_train,y_train)
```

```python
y_pred = log_model.predict(scaled_x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix ,classification_report , plot_confusion_matrix
```

```python
# Accuracy can be a useful measure if we have the same amount of samples per class but if we have an imbalanced set of samples accuracy isn't useful at all. Even more so, a test can have a high accuracy but actually perform worse than a test with a lower accuracy.
accuracy_score(y_test , y_pred)
```

```python
confusion_matrix(y_test , y_pred)
```

```python
plot_confusion_matrix(log_model , scaled_x_test ,y_test)
```

```python
from sklearn.metrics import plot_precision_recall_curve , plot_roc_curve
```

```python
plot_roc_curve(log_model ,scaled_x_test , y_test)
```

```python
plot_precision_recall_curve(log_model , scaled_x_test , y_test)
```
