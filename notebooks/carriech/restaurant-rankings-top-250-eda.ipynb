```python
import numpy as np
import pandas as pd
import plotly.express as px
```

**<font color = MediumVioletRed size = 5>Load data</font>**

```python
top250 = pd.read_csv('/kaggle/input/restaurant-business-rankings-2020/Top250.csv')
```

**<font color = MediumVioletRed size = 5>Show first five row of data</font>**

```python
top250.head()
```

**<font color = MediumVioletRed size = 5>Information about data</font>**

```python
top250.info()
```

**<font color = MediumVioletRed size = 5>Data cleaning</font>**

```python
top250.isnull().sum()
```

```python
top250.drop(['Content','Headquarters'],axis = 1,inplace = True)
top250.columns
```

```python
df1 = top250.groupby('Segment_Category').sum().reset_index().sort_values('Sales', ascending = False).reset_index().head(10)
fig = px.pie(df1 , values = 'Sales', names = 'Segment_Category',
            color_discrete_sequence = px.colors.sequential.Rainbow)

fig.update_layout(title='Top 10 - Sales Based on the Segment Category',
                  title_x = 0.4,
                  title_font = dict(size = 16, color = 'MidnightBlue')
                 )

fig.show()
```

```python
splitcategory = top250["Segment_Category"].str.split("&", n = -1,  expand = True)
```

```python
top250['category'] = splitcategory[0]
```

```python
top250['category'].unique()
```

```python
df1 = top250.groupby('category').sum().reset_index().sort_values('Sales', ascending = False).reset_index().head(10)
fig = px.pie(df1, values = 'Sales', names = 'category',color_discrete_sequence=px.colors.qualitative.T10)

fig.update_layout(title='Top 10 - Sales Based on the Category',
                  title_x = 0.45,
                  title_font = dict(size = 16, color = 'MidnightBlue')
                 
                 )

fig.show()
```

```python
df1 = top250.groupby("Restaurant").sum().reset_index().sort_values('Sales', ascending=False).reset_index().head(10)

fig = px.bar(df1, x = 'Restaurant', y = 'Sales', color = 'Sales', color_continuous_scale = 'ice')

fig.update_layout(title=' Top 10 - Sales Based on the Restaurant',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'MidnightBlue'))

fig.show()
```

```python
top250['YOY_Sales']=top250['YOY_Sales'].str.strip('%')
top250['YOY_Sales']=top250['YOY_Sales'].astype(float)
```

```python
df1 = top250.groupby('Segment_Category').sum().reset_index().sort_values('YOY_Sales',ascending = False).reset_index().head(10)

fig = px.pie(df1,values='YOY_Sales', names='Segment_Category',color_discrete_sequence=px.colors.sequential.RdBu)
fig.update_traces(textposition = 'inside',textinfo = 'percent+label')
fig.update_layout(title='Top 10 - Year over Year increase in  Sales Based on the Segment Category',
                  title_x = 0.4,
                  title_font = dict(size = 16, color = 'Darkred'))

fig.show()
```

```python
yoy_sales_postive=top250[top250['YOY_Sales']>0]
yoy_sales_negative=top250[top250['YOY_Sales']<0]
```

```python
df1 = yoy_sales_postive.groupby("Restaurant").sum().reset_index().sort_values('YOY_Sales',ascending = False).reset_index().head(10)

fig = px.bar(df1, y = 'YOY_Sales', x = 'Restaurant',color = 'YOY_Sales',
             color_continuous_scale = 'blues')

fig.update_layout(title=' Top 10 - Year over Year increase in Sales Based on the Restaurant',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'MidnightBlue'))

fig.show()
```

```python
df1 =yoy_sales_negative.groupby("Restaurant").sum().reset_index().sort_values('YOY_Sales',ascending = True).reset_index().head(10)

fig = px.bar(df1, y = 'YOY_Sales',x = 'Restaurant',color='YOY_Sales',
             hover_data = ['Restaurant', 'YOY_Sales'],
             color_continuous_scale = 'ice')

fig.update_layout(title='Bottom 10 - Year over Year Decrease in Sales Based on the Restaurant',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'MidnightBlue'))

fig.show()
```

```python
top250['YOY_Units']=top250['YOY_Units'].str.strip('%')
top250['YOY_Units']=top250['YOY_Units'].astype(float)
```

```python
yoy_units_postive=top250[top250['YOY_Units']>0]
yoy_units_negative=top250[top250['YOY_Units']<0]
```

```python
df1 = top250.groupby('Segment_Category').sum().reset_index().sort_values('YOY_Units', ascending=False).reset_index().head(10)

fig = px.pie(df1, values='YOY_Units', names='Segment_Category',template="seaborn")

fig.update_traces(textposition='inside', textinfo='value')
fig.update_layout(title=' Top 10 - Year over Year increase in Units Based on the Segment Category',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'MidnightBlue'))
fig.show()
```

```python
df1 = yoy_units_postive.groupby("Restaurant").sum().reset_index().sort_values('YOY_Units',ascending = False).reset_index().head(10)

fig = px.bar(df1, y = 'YOY_Units',x = 'Restaurant', color = 'YOY_Units',
             color_continuous_scale = 'brbg')

fig.update_layout(title='Top 10 - Year over Year increase in Units Based on the Restaurant',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'Teal'))

fig.show()
```

```python
df1 =yoy_units_negative.groupby("Restaurant").sum().reset_index().sort_values('YOY_Units',ascending = True).reset_index().head(10)

fig = px.bar(df1, y = 'YOY_Units',x = 'Restaurant',color='YOY_Units',
             hover_data = ['Restaurant', 'YOY_Units'],
             color_continuous_scale = 'ice',
             height = 400
            )

fig.update_layout(title=' Bottom 10 - Year over Year Decrease in Units Based on the Restaurant',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'MidnightBlue'))

fig.show()
```
