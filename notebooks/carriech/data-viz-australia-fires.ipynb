```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("../input/fires-from-space-australia-and-new-zeland/fire_nrt_M6_96619.csv")
```

```python
df1 = df.sort_values(by = ['acq_date'],ascending = True)

fig = px.density_mapbox(df1, lon = 'longitude', lat = 'latitude',z = 'brightness', radius = 7,
                        center = dict(lon = 140,lat = -25),zoom = 2.7,
                        mapbox_style = 'carto-positron',animation_frame = 'acq_date',
                        color_continuous_scale = 'redor')


fig.update_layout(title = 'Australia Fires: 2019/10/01 to 2020/01/11',
                  title_font = dict(size = 22, family = 'Balto',color = 'FireBrick'),
                  title_x = 0.5)
fig.show()
```
