```python
df1 = df.iloc[:50,:]

tr1 = go.Scatter(x = df1.world_rank,
                 y = df1.citations,
                 mode = 'lines+markers',
                 name = 'Citation',
                 marker = dict(color = 'LightSteelBlue'),
                 text = df1.university_name)

tr2 = go.Scatter(x = df1.world_rank,
                 y = df1.teaching,
                 mode = 'lines+markers',
                 name = 'Teaching',
                 marker = dict(color = 'SteelBlue'),
                 text = df1.university_name)

data = [tr1,tr2]
layout = dict(title = 'Top 50 university - Citation and Teaching',
              title_x = 0.5,
              template = "plotly_white",
              title_font = dict(size = 16, color = 'DarkBlue'),
              xaxis = dict(title = 'World Rank'))

fig = go.Figure(data = data, layout = layout)
fig.show()
```

```python
df= df.dropna(subset = ['num_students','international_students'])
```

```python
df[['num_students','international_students']].isnull().sum()
```

```python
df2016 = df[df.year==2016].iloc[:10,:]

pie1 = [float(each.replace(',','.')) for each in df2016.num_students]
labels = df2016.university_name

fig = px.pie(df2016, values = pie1, 
             names = labels,
             hole = 0.3,
             labels = {'university_name':'University Name'},
             color_discrete_sequence = px.colors.sequential.GnBu
            
            )

fig.update_layout(title = 'Top 10 university - The number of student (2016)',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'DarkBlue'),
                  annotations = [dict(font = dict(size = 10,color = 'MediumBlue'),
                                      showarrow = False,
                                      text = 'Number of Student',
                                      x = 0.5, 
                                      y = 0.5)])

fig.show()
```

```python
df = df[df['student_staff_ratio'].notna()]
```

```python
df['student_staff_ratio'].isnull().sum()
```

```python
df2014 = df[df.year==2014]
df2015 = df[df.year==2015]
df2016 = df[df.year==2016]

tr1 = go.Histogram(x = df2014.student_staff_ratio,
                   opacity = 0.7,
                   name = '2014',
                   marker = dict(color = 'SteelBlue'))

tr2 = go.Histogram(x = df2015.student_staff_ratio,
                   opacity = 0.7,
                   name = '2015',
                   marker = dict(color = 'LightBlue'))

tr3 = go.Histogram(x = df2016.student_staff_ratio,
                   opacity = 0.7,
                   name = '2016',
                   marker = dict(color = 'BlanchedAlmond'))

data = [tr1,tr2,tr3]

layout = dict(title = ' The Ratio of Student-staff (2014,2015,2016)',
              title_font = dict(size = 16, color = 'DarkBlue'),
              title_x = 0.5,
              barmode = 'overlay',
              xaxis = dict(title = 'Ratio'),
              yaxis = dict(title = 'Count'),
              template = 'plotly_white',
             
             )

fig = go.Figure(data = data, layout = layout)
fig.show()
```

```python
dfcountry = df['country'].value_counts().reset_index().head(20)
fig = px.bar(dfcountry, 
             x = 'index', 
             y = 'country', 
             color = 'country',
             color_continuous_scale = 'gnbu',
             labels = {'index':'Country','country':'Count'})

fig.update_layout(title = 'Top 20 Countries with the most university',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'DarkBlue'))
fig.show()
```
