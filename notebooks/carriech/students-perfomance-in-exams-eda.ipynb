**<font color = Teal size = 5>Data Analysis</font>**

```python
px.pie(df,df['gender'])
```

**The number of female is more than males.**

```python
#Count plot of gender, race, level of eductaion and test preparation
plt.figure(figsize = [10,7])
fig,axes = plt.subplots(2,2, figsize = [18,10])
sns.countplot(x = df['gender'],ax = axes[0,0])
sns.countplot(x = df['race/ethnicity'],ax = axes[0,1],palette = 'deep')
sns.countplot(x = df['parental level of education'],ax = axes[1,0])
sns.countplot(x = df['test preparation course'],ax = axes[1,1],palette = 'dark')
```

```python
pd.crosstab(df['gender'],df['test preparation course']).plot(kind = 'barh')
plt.show()
```

**The number of students are male and female who completed test preparation course is less than those who did not complete.**

```python
df['total'] = (df['math score'] + df['reading score'] + df['writing score']) / 3
sns.barplot(x = df['test preparation course'], y = df['total'], palette = 'crest')
```

**Students who completed test preparation courses get higher score.**

```python
df.groupby('gender')['math score','reading score','writing score'].mean().plot(kind = 'bar')
plt.show()
```

* **Male score in math is higher than female.**
* **Female score in reading and writing are higher than male.**

```python
pd.crosstab(df['gender'],df["race/ethnicity"]).plot(kind = 'bar')
plt.show()
```

* **The number of male in group A is more than famales.**
* **The number of famale in group B and group C are more than males.**
* **The number of male and famale in group D are probably the same in group E.**

```python
pd.crosstab(df["parental level of education"],df["race/ethnicity"]).plot(kind = 'bar')
plt.show()
```

```python
df['total'] = (df['math score'] + df['reading score'] + df['writing score']) / 3
sns.barplot(y = df['parental level of education'], x = df['total'],palette = 'OrRd')
```

**The score of student whose parents possess master and bachelor level education are higher than others.**

```python
df1 =df['math score'].value_counts().reset_index()
df2 =df['reading score'].value_counts().reset_index()
df3 =df['writing score'].value_counts().reset_index()

tr1 = go.Bar(x = df1['index'],
             y = df1['math score'],
             name = 'math',
             xaxis = 'x',
             yaxis = 'y',
             marker = dict(color = 'rgb(255, 20, 147)'))

tr2 = go.Bar(x = df2['index'],
             y = df2['reading score'],
             name = 'reading',
             xaxis = 'x2',
             yaxis = 'y2',
             marker = dict(color = 'rgb(255, 69, 0)'))

tr3 = go.Bar(x = df3['index'],
             y = df3['writing score'],
             name = 'writing',
             xaxis = 'x3',
             yaxis = 'y3',
             marker = dict(color = 'rgb(240, 230, 140)'))

data = [tr1,tr2,tr3]

layout = go.Layout(xaxis = dict(domain=[0, 0.45],anchor = 'y'),
                   yaxis = dict(domain=[0, 0.45],anchor = 'x'),
                   xaxis2 = dict(domain=[0.55, 1],anchor = 'y2'),
                   yaxis2 = dict(domain=[0, 0.45],anchor = 'x2'),
                   xaxis3 = dict(domain=[0, 0.45],anchor = 'y3'),
                   yaxis3 = dict(domain=[0.55, 1],anchor  = 'x3'))

fig = go.Figure(data =data,layout =layout)
fig.show()
```

**Reading and wirting score is higher than math.**

```python
df['total'] = (df['math score'] + df['reading score'] + df['writing score']) / 3
sns.barplot(x = df['lunch'], y = df['total'],palette = 'bwr')
```

**The score of students who were provided standard lunch is higher than those who were provided free/reduced lunch.**

```python
plt.figure(figsize = [10,7])
fig,axes = plt.subplots(1,3, figsize = [18,10])
sns.barplot(data = df, x='race/ethnicity', y='math score',ax = axes[0],palette = 'bwr')
sns.barplot(data = df, x='race/ethnicity', y='reading score',ax = axes[1],palette = 'deep')
sns.barplot(data = df, x='race/ethnicity', y='writing score',ax = axes[2],palette = 'PuBu')
```

**These bars show that students in group E perform the best score in all subjects, while group A perform the worst score.**

```python
plt.figure(figsize = [10,7])
fig,axes = plt.subplots(2,2, figsize = [18,10])
sns.scatterplot(x = 'math score',y = 'reading score',hue = 'parental level of education',data = df, ax =axes[0,0],palette = 'deep')
sns.scatterplot(x = 'math score',y = 'reading score',hue = 'race/ethnicity',data = df, ax =axes[0,1],palette = 'bright')
sns.scatterplot(x = 'math score',y = 'reading score',hue = 'lunch',data = df, ax =axes[1,0],palette = 'dark')
sns.scatterplot(x = 'math score',y = 'reading score',hue = 'test preparation course',data = df, ax =axes[1,1],palette = 'pastel')
```

**<font color = Teal size = 5>Conclusion</font>**

* **Students who completed test preparation courses get higher score.**
* **The score of female in reading and writing is higher than male, but the score of male in math is higher than female.**
* **The score of student whose parents possess master and bachelor level education are higher than others.**
* **The score of students who were provided standard lunch is higher than those who were provided free/reduced lunch.**
* **The students in group E perform the best score in all subjects, while those in group A perform the worst score.**
