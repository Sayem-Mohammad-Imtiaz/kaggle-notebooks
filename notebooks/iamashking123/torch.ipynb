```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('../input/tmdb-movie-metadata/tmdb_5000_movies.csv',usecols=['id','title','overview'])
```

```python
df['overview'] = df['overview'].fillna('')
```

```python
userinput = 'userinput'
detail = 'tony stark iron man'
cid = 171122223123
```

```python
df.loc[len(df.index)] = [cid, detail, userinput] 
```

```python
tfidf = TfidfVectorizer(stop_words='english')
mat = tfidf.fit_transform(df['overview'])
```

```python
cos_sim = cosine_similarity(mat,mat)
```

```python
title2id = pd.Series(df.index,index=df['title'])
```

```python
def recommend(title,measure,npreds):
    idx = title2id[title]
    score = list(enumerate(measure[idx]))
    score = sorted(score,key=lambda x: x[1],reverse=True)
    score = score[1:npreds]
    idxs = [i[0] for i in score]
    return df['title'].iloc[idxs]
```

```python
recommend(userinput,cos_sim,20)
```
