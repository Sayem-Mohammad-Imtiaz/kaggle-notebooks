```python
import pandas as pd
```

```python
df=pd.read_csv('../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
x=df.drop("DEATH_EVENT",axis=1)
x
```

```python
y=df['DEATH_EVENT']
y
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data visualization

```python
plt.figure(figsize=(25,25))
sns.heatmap(df.corr(),annot=True,cmap="RdYlGn")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=2)
```

Feature scaling

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x_train=scale.fit_transform(x_train)
x_test=scale.fit_transform(x_test)
```

# Random Forest Classifier

```python
#Finding the optimum number of n_estimators

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
list1 = []
for estimators in range(10,30):
    classifier = RandomForestClassifier(n_estimators = estimators, random_state=0, criterion='entropy')
    classifier.fit(x_train, y_train)
    y_pred = classifier.predict(x_test)
    list1.append(accuracy_score(y_test,y_pred))
#print(mylist)
plt.plot(list(range(10,30)), list1)
plt.show()
```

```python
# Training the RandomForest Classifier on the Training set

from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 26, criterion='entropy', random_state=2)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

# Accuracy Testing

```python
from sklearn import metrics
```

```python
print("ACCURACY OF THE MODEL: ", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
mylist=[]
con_mat = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
mylist.append(accuracy)
print(con_mat)
print(accuracy)
```

# Decision Tree Classifier

```python
#Finding the optimum number of max leaf nodes

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
list1=[]
for leaves in range(2,10):
    classifier=DecisionTreeClassifier(max_leaf_nodes=leaves,
                                      random_state=0,criterion='entropy')
    classifier.fit(x_train,y_train)
    y_pred=classifier.predict(x_test)
    list1.append(accuracy_score(y_test,y_pred))
plt.plot(list(range(2,10)),list1)
plt.show()
```

```python
# Training the Decision Tree Classifier on the Training set

classifier = DecisionTreeClassifier(max_leaf_nodes = 3, random_state=0, criterion='entropy')
classifier.fit(x_train, y_train)
```

```python
# Predicting the test set results

y_pred = classifier.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
ac = accuracy_score(y_test, y_pred)
mylist.append(ac)
print(cm)
print(ac)
```
