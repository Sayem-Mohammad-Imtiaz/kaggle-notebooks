*First we'll load our data*

```python
dataset = pd.read_csv("../input/water-potability/water_potability.csv")
dataset.head()
```

Now we'll find the sum of all null values in our data

```python
print(dataset.isnull().sum())
```

> Use the mean values of each column in the csv to fill all null values

```python
dataset["ph"] = dataset["ph"].fillna(dataset["ph"].mean())
dataset["Sulfate"] = dataset["Sulfate"].fillna(dataset["Sulfate"].mean())
dataset["Trihalomethanes"] = dataset["Trihalomethanes"].fillna(dataset["Trihalomethanes"].mean())
```

Now check all the datasets for null values again 

```python
print(dataset.isnull().sum())
```

So we have no null values. That means we can proceed with out process.

Now we are gonna plot our values and visualize it

First we'll see how much samples in our dataset is portable

```python
%matplotlib inline
sns.histplot(dataset["Potability"], color='cyan')
plt.show()
```

It shows that there are 1278 values with Potability 1 and 1998 values with Potability 0

# Data Visualization

```python
ph = dataset["ph"]
potability = dataset["Potability"]
sulfate = dataset["Sulfate"]
hardness = dataset["Hardness"]
solids = dataset["Solids"]
conductivity = dataset["Conductivity"]
organic_carbon = dataset["Organic_carbon"]
trihalomethanes = dataset["Trihalomethanes"]
chloramines = dataset["Chloramines"]
turbidity = dataset["Turbidity"]
```

```python
# Plotting ph
%matplotlib inline
plt.figsize = (20,6)
sns.kdeplot(data=dataset, x="ph",y="Hardness", hue="Potability", fill=True)
plt.show()
```

The water is potable if it contains less solids and the pH value in the range 6 to 8

Lets now mark for sulphates and Chloramines

```python
sns.kdeplot(data=dataset, x="Solids", y="Chloramines", hue="Potability", fill=True)
plt.show()
```

In the above chart water with chloramine values 6 to 8 and with less solid content are ideal

```python
sns.scatterplot(data=dataset, x="Sulfate", y="Conductivity", hue="Potability")
plt.show()
```

Water with low conductivity and low sulphate are ideal(Potable)
