```python
from tensorflow.keras.models import *
from tensorflow.keras.applications import VGG16
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.layers import *
from tensorflow.keras.activations import  sigmoid
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.optimizers import *
from tensorflow.keras.callbacks import *
```

```python
def embedding(inp):
     x = Embedding(output_dim=10, input_dim=2, input_length=1)(inp)
     x = Flatten()(x)
     return Dense(10, activation='relu')(x)
```

```python
def nn_model():
    x_data = Input(shape=(12,),name="data")
#     cats_in = [Input(shape=(1,),name=cat) for cat in ['anaemia','diabetes','high_blood_pressure','sex','smoking']]
#     cats = concatenate([embedding(cat) for cat in cats_in])
    x = sigmoid(x_data)
    x = Dense(128,activation='relu')(x)
    
    
    
    x = Dense(64,activation='relu')(x)
    
    x = Dense(1,activation='sigmoid')(x)
    
    return Model(inputs = x_data ,outputs = x)
```

```python
model = nn_model()
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2,
                              patience=2, min_lr=0.00001)
model.fit(X_train,y_train,epochs=100,callbacks=[reduce_lr],batch_size=1)
```

```python
model.evaluate(X_test,y_test)
```
