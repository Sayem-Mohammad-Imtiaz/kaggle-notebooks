```python
# Tarefa: Leia o CSV com o Pandas
# Dataset: Review = textos com uma review
#          Sentiment = sentimento (se é uma review positiva ou negativa)

reviews = pandas.read_csv("../input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv")
reviews
```

```python
# Tarefa: Divida o X e Y e use o train_test_split
X = reviews["review"][0:5000]
Y = reviews["sentiment"][0:5000]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y)
```

```python
# Tarefa: Use o vetorizador CountVectorizer para fazer o fit_transform e transform
Vectorizer = CountVectorizer()
X_train_vectorized = Vectorizer.fit_transform(X_train)
X_test_vectorized = Vectorizer.transform(X_test)
```

```python
# Tarefa: Use a regressão logística

model = LogisticRegression(max_iter=3000)
model.fit(X_train_vectorized, Y_train)

resultado = model.predict(X_test_vectorized)
print(accuracy_score(resultado, Y_test))
print(confusion_matrix(resultado, Y_test))
```

```python
# Tarefa: Teste um exemplo negativo e um positivo

nova_review = Vectorizer.transform(["This movie is bad", "This movie is awesome"])
model.predict(nova_review)
```

```python
# Vamos descombrir por que deu o warning na célula passada?

from sklearn.model_selection import learning_curve
def plot_learning_curve(est, X_train, y_train) :
    train_sizes, train_scores, test_scores = learning_curve(estimator=est, X=X_train, y=y_train,cv=5,n_jobs=-1)
    train_mean = numpy.mean(train_scores, axis=1)
    train_std = numpy.std(train_scores, axis=1)
    test_mean = numpy.mean(test_scores, axis=1)
    test_std = numpy.std(test_scores, axis=1)
    matplotlib.figure(figsize=(8,10))
    matplotlib.plot(train_sizes, train_mean, color='blue', marker='o', markersize=5, label='training accuracy')
    matplotlib.fill_between(train_sizes, train_mean + train_std, train_mean - train_std, alpha=0.15, color='blue')
    matplotlib.plot(train_sizes, test_mean,color='green', linestyle='--',marker='s', markersize=5,label='validation accuracy')
    matplotlib.fill_between(train_sizes,test_mean + test_std,test_mean - test_std,alpha=0.15, color='green')
    matplotlib.grid(b='on')
    matplotlib.xlabel('Number of training samples')
    matplotlib.ylabel('Accuracy')
    matplotlib.legend(loc='lower right')
    matplotlib.show()

plot_learning_curve(model, X_train_vectorized, Y_train)
```
