```python
frame["genres"][0].split("|")
unique_genres = []
for i in range(0, 5043):
    g = frame["genres"][i].split("|")
    for i in g:
        if i not in unique_genres:
            unique_genres.append(i)
print(len(unique_genres))
print(unique_genres)
```
