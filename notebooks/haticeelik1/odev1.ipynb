```python
#nested func. içiçe fonksiyonlar
def once():
    def sonra():
        x=6
        y=x+2
        return y
    return sonra()*3
print(once())
```

```python
#default ve flexible func
def f(c,h=1,t=0):
    x=(h*t)+c
    return x
print(f(3))
print(f(3,2,1))
print(" ")
def f(*args):
    for i in args:
        print(i)
f(3)
print(" ")
f(1,2,3,4,4)
print(" ")
def f(**kwargs):
    for key,value in kwargs.items():
        print(key," : ",value)
print(f(ilk = 'a', iki = 'b'))
```

```python
#lambda func
carp = lambda x: x*3
top = lambda a,b,c: (a*b)+c
print(carp(4))
print(top(3,0,1))
```

```python
#map func
liste = [3,4,5]
orn = map(lambda x: x**2,liste)
print(list(orn))
```

```python
#tidy data
data_yeni = data.head()
data_yeni
```

```python
melted = pd.melt(frame=data_yeni,id_vars='Country or region',value_vars=['Score','Generosity'])
melted
#frame=hangi data setini kullanmak istediğini belirtirsin,id_vars= hangi sütunun melt edilmemesini istersen,value_vars=hangi sütunlar melt edilsin
```

```python
#pivoting data melt edilen verileri eski haline getirmek için
melted.pivot(index='Country or region',columns='variable',values='value')
#index=yukarıdaki id_vars yerine yazdığın
#columns yukarıdaki tabloda variable sütunundakiler
#values tablodaki value değerleri
```

```python
#concatenating data
data1=data.head()
data2=data.tail()
concatdata = pd.concat([data1,data2],axis = 0, ignore_index=True)
concatdata
#axis=0 altalta birleştirmek için
#ignore_index=True o verilerin indexlerini görmezden gel yeni indexler ata demek
```

```python
#bu tabloda hiç NaN değeri yok ama olanlarda bu eksikliği düzeltmek gerekmektedir.Bunun için
data['Generosity'].value_counts(dropna=False)
#burada hangi değerden kaç tane var saydırdık ve hiç NaN değeri yokmuş.Eğer olsaydı bunu gidermek için
```
