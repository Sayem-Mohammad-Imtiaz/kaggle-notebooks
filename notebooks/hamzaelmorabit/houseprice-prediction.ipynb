```python

import numpy as np
import pandas as pd 
import math
```

```python
df = pd.read_csv('../input/housesalesprediction/kc_house_data.csv')

df = df[ ['sqft_lot', 'floors', 'sqft_living','price' , ] ]

print(np.isnan(df).sum())

#80% des données apprentissage
#20% des données pour le test
train_df = df[0:17000]
# print(train_df )

test_df  = df[17000:]
# train_df
# test_df
# df['log_sqft_lot'].hist(bins=100)
```

## z-score normalize 

```python
                            
A = train_df
train_df = (A - A.mean(axis=0)) / A.std(axis=0)
# print(train_df)
# A.hist(bins=50)

A = test_df
test_df = (A - A.mean(axis=0)) / A.std(axis=0)
# A.hist(bins=50)
# print(test_df)
```

```python
train_X = train_df[ ['sqft_lot', 'floors','sqft_living'] ].values
train_Y = train_df[ 'price' ].values
# print(train_X)

test_X = test_df[ ['sqft_lot', 'floors','sqft_living'] ].values
test_Y = test_df[ 'price' ].values

# print( train_X.shape, test_X.shape)
# print( train_Y.shape, test_Y.shape )
```

```python
import keras
from keras import layers
from keras.models import Model
from keras.optimizers import SGD
```

```python


inpt = layers.Input( shape=(3,) )
x    = layers.Dense(200) ( inpt )
actv = layers.Activation('relu') (x)
# x    = layers.Dense(200) ( actv )
# actv = layers.Activation('relu') (x)
output = layers.Dense(1) (actv)

model = Model( inpt, output )
model.summary()
```

## Early Stopping in Keras 

```python

from keras import callbacks 
earlystopping = callbacks.EarlyStopping(monitor ="val_loss",  
                                        mode ="min", patience = 5,  
                                        restore_best_weights = True) 
  
```

```python
opt = SGD() 

model.compile( optimizer=opt, loss='mse')
```

```python


model.fit( train_X, train_Y, epochs=80 ,validation_data = (test_X,test_Y),  callbacks =[earlystopping]) 
```
