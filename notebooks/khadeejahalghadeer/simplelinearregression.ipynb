In this leaaon, I am showing Simple linear regression Machine learning model for beginner.
there are two variables. Independent variable:YearsExperience and dependent variable(target variable):Salary. Linear regression model will predict the salary based on the years of experience \

1. Read the Data
2. Plot the Data 
3. Defined indepenent and target variables
4. fit linear Regression model then predict the value
5. Save the model 


```python
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/kaggle/input/salary-data-simple-linear-regression/Salary_Data.csv')
df.head()
```

```python
plt.scatter(df.YearsExperience,df.Salary)
plt.title('Experience vs Salary')
plt.xlabel(' Experience(years)')
plt.ylabel('Salay')
```

```python
X = df.drop('Salary',axis='columns')
X.head(5)
```

```python
y= df.Salary
y.head(5)
```

```python
reg = linear_model.LinearRegression()
reg.fit(X,y)
```

```python
reg.predict([[3]])
```

```python
#save the  model
import pickle
pickle.dump(reg,open('model.pkl','wb'))
```

```python
#load the model
model= pickle.load(open('model.pkl','rb'))
model.predict([[4]])
```

```python
print(model.predict([[1.8]]))
```
