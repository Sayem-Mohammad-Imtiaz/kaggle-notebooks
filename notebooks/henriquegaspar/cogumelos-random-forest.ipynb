```python
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score
from yellowbrick.classifier import ConfusionMatrix
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
import seaborn as srs
```

```python
cogumelos = pd.read_csv("../input/mushroom-classification/mushrooms.csv")
```

```python
cogumelos.head()
```

```python
cogumelos = cogumelos.apply(LabelEncoder().fit_transform)
```

```python
cogumelos.head()
```

```python
plt.figure(figsize=(16,8))
srs.heatmap(cogumelos.corr(), annot=True)
```

```python
treino = cogumelos.iloc[:,1:23]
teste = cogumelos.iloc[:,0]
```

```python
x_teste, x_treino, y_teste, y_treino = train_test_split(treino,
                                                       teste,
                                                        test_size=0.3,
                                                       random_state=0)
```

```python
modelo = RandomForestClassifier(n_estimators=100)
modelo.fit(x_treino, y_treino)
```

```python
previsoes = modelo.predict(x_teste)
confusao = confusion_matrix(y_teste, previsoes)
confusao
```

```python
print(classification_report(y_teste, previsoes))
```

```python
print(accuracy_score(y_teste, previsoes))
```
