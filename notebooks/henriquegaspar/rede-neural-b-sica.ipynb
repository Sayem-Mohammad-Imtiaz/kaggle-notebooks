```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
data = pd.read_csv("../input/breast-cancer-wisconsin-data/data.csv")
```

```python
data.head()
```

```python
data['diagnosis'].value_counts()
```

```python
encoder = LabelEncoder()
data['diagnosis'] = encoder.fit_transform(data['diagnosis'])
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
entradas = data.drop('diagnosis', axis=1)
saidas = pd.DataFrame(data['diagnosis'])
```

#### Separação da base de dados em treino e teste ####

```python
entrada_treino, entrada_teste, saida_treino, saida_teste = train_test_split(entradas, saidas, test_size=0.25)
```

```python
from keras.models import Sequential
from keras.layers import Dense
import keras
```

#### Construção da Rede Neural ####

```python
rede_neural = Sequential()
#Para determinar a variavél units, basta aplicar a formula (variavel.shape[1] + possiveis saídas / 2)
#No caso, (entrada_treino.shape[1] + 1 / 2) = 15.5, arredonda-se para 16.
#Input_dim apenas na primeira entrada.
rede_neural.add(Dense(units=16, activation='relu', kernel_initializer='random_uniform', input_dim=31))
rede_neural.add(Dense(units=1, activation='sigmoid'))
#Um learning rate elevado é mais rápido, e um lr menor é mais rápido, mas tende a não chegar no mínimo gradiente.
#Assim, aplica-se uma taxa de decay, que é subtraído em cada etapa que o lr percorre no gradiente.
otimizador = keras.optimizers.Adam(learning_rate = 0.001, decay = 0.0001, clipvalue = 0.5)
rede_neural.compile(optimizer=otimizador, loss='binary_crossentropy', metrics=['binary_accuracy'])
rede_neural.fit(entrada_treino, saida_treino, batch_size=10, epochs=100)
```

##### Visualização de pesos #####

```python
pesos = rede_neural.layers[0].get_weights()
```

##### Predição #####

```python
previsoes = rede_neural.predict(entrada_teste)
previsoes = (previsoes>0.5)
```

#### Aplicação das métricas de acerto para avaliação da acurácia da rede neural ####

```python
from sklearn.metrics import classification_report, confusion_matrix

print(classification_report(saida_teste, previsoes))
```

```python
print(confusion_matrix(saida_teste, previsoes))
```

```python
print(rede_neural.evaluate(entrada_teste, saida_teste))
```
