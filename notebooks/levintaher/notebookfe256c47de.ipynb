```python
import numpy as np
import pandas as pd
import datetime
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from wordcloud import STOPWORDS, WordCloud
import pylab as plt
from matplotlib.pyplot import figure
%matplotlib inline
from plotly import __version__
from plotly.offline import download_plotlyjs, plot, iplot
import cufflinks as cf

        
npr = pd.read_csv("../input/imdb-extensive-dataset/IMDb movies.csv", dtype={"year": "string"})

npr = npr[(npr['year'] >= '2016') & (npr['year'] <= '2016')]

npr = npr[(npr['country'] == 'USA')]
```

```python
cv = CountVectorizer(max_df=0.02, min_df=0, stop_words=STOPWORDS)
```

```python
dtm = cv.fit_transform(npr['description'].fillna(''))
```

```python
dtm
```

```python
from sklearn.decomposition import LatentDirichletAllocation

LDA = LatentDirichletAllocation(n_components=20,random_state=42)

LDA.fit(dtm)
```

# Die Begriffe eines Themas

Hier hat man einen groben Überklick über die Begriffe eines Themas. Es kann helfen die Modellierung einen bestimmten Thema zuzuordnen. In dem Beispiel fallen Begriffe wie "Relationship", "Gay", "Cross", "America", "Society" etc. auf, die auf Themen wie gleichgeschlechtliche Beziehungen oder gesellschaftlicher Wandel deuten können.

```python
single_topic = LDA.components_[6]
top_word_indices = single_topic.argsort()[-120:]
arr = np.array(top_word_indices)
reversed_arr = arr[::-1]

wordsFromTopic = ' '.join(cv.get_feature_names()[index] for index in reversed_arr)
topic_wordcloud = WordCloud(stopwords=STOPWORDS, background_color='white', height=2000, width=4000).generate(wordsFromTopic)
plt.figure(figsize=(14,8))
plt.imshow(topic_wordcloud)
plt.axis('off')
plt.show()
```

```python
topic_results = LDA.transform(dtm)
```

```python
topic_results.shape
```

```python
npr['Thema'] = topic_results.argmax(axis=1)
```

```python
tname = npr['Thema'].value_counts()
#tname['test'] = tname.pop(1)

print(tname)
```
