```python
import plotly.offline as py
import plotly.graph_objs as go
import plotly.tools as tls
from plotly.offline import init_notebook_mode, iplot, plot
```

```python
df1 = df_train[df_train.target == 0].iloc[:,:]
df2 = df_train[df_train.target == 1].iloc[:,:]

# creating trace1
trace1 =go.Scatter(
                    x = df1.slope,
                    y = df1.oldpeak,
                    mode = "markers",
                    name = "No Heart Disease",
                    marker = dict(color = 'rgba(255, 128, 255, 1.5)')                   
                    )
# creating trace2
trace2 =go.Scatter(
                    x = df2.slope,
                    y = df2.oldpeak,
                    mode = "markers",
                    name = "Heart Disease",
                    marker = dict(color = 'rgba(0, 128, 255, 1.5)')                    
                    )


data = [trace1, trace2]

layout = dict(
              xaxis= dict(title= 'Slope',ticklen= 3,zeroline= False),
              yaxis= dict(title= 'oldpeak',ticklen= 3,zeroline= False)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```

```python
trace1 = go.Scatter3d(
    x=df_train.target,
    y=df_train.slope,
    z=df_train.oldpeak,
    mode='markers',
    marker=dict(
        size=10,
        color='rgb(255,128,128)',                # set color to an array/list of desired values      
    )
)

data = [trace1]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0  
    )
    
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
df = df_train[df_train.target == 1].iloc[:,:]

x = df.slope

trace1 = {
  'x': x,
  'y': df.oldpeak,
  'name': 'oldpeak',
  'type': 'bar'
};
trace2 = {
  'x': x,
  'y': df.oldpeakBand,
  'name': 'oldpeakBand',
  'type': 'bar'
};
data = [trace1, trace2];
layout = {
  'xaxis': {'title': 'Slope and Oldpeak'},
  'barmode': 'relative'
};
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```
