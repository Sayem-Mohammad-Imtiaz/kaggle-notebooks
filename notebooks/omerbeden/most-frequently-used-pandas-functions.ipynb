# Dealing with Missing Values

* We can see which columns have nan/null value in many ways.

```python
covid_df.isna().any()
```

```python
covid_df.isnull().any()
```

* we can also obtain Nan/null values getting notnull values like this 

```python
covid_df.notnull().all()
```

* we can fill this missing values with a skalar or statistical methods

```python
covid_df.fillna(value=0)
```

* or we can drop null values
