```python
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
X = data
scaler = MinMaxScaler().fit(X)
X = scaler.transform(X)
X = pd.DataFrame(X,columns= data.columns)
```

```python
X.corr()['Churn'].sort_values(ascending=False)
```

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import f1_score
Y = X['Churn'].astype('int')
X = X.drop(columns=['Churn'])

oversample = SMOTE()
X_smote, Y_smote = oversample.fit_resample(X, Y)

rf = RandomForestClassifier()
gbc = GradientBoostingClassifier()

```

```python
def train(X,Y):
    X_train,X_test,y_train,y_ytest = train_test_split(X,Y,test_size=0.3,random_state=42)
    len(X_train),len(X_test)
    rf.fit(X_train,y_train)
    #gbc.fit(X_train,y_train)
    rf.score(X_test,y_ytest)
    plot_confusion_matrix(rf,X_test,y_ytest)
    #plot_confusion_matrix(gbc,X_test,y_ytest)
    y_pred1 = rf.predict(X_test)
   # y_pred2 = gbc.predict(X_test)
    f11 = f1_score(y_ytest,y_pred1)
    #f12 = f1_score(y_ytest,y_pred2)
    return f11
    
```

```python
f1 = train(X,Y)
f2 = train(X_smote,Y_smote)
```

```python
print("Original   : ",f1)
print("With Smote : ",f2)
```
