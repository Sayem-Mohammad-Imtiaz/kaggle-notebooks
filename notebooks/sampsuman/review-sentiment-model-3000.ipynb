```python
data=pd.read_csv('/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv')
```

```python
import pickle
import re
# Counter Library
from collections import Counter
```

```python
word_dict=pickle.load(open('/kaggle/input/most-common-imdb-review-words-full-dictionary/word_dict.pkl','rb'))
```

```python
word_list=word_dict.most_common(3000)
```

```python
word_list=word_list[15:]
```

```python
# For most_common-2985 words
features=[]
labels=[]
for i in range(0,data.shape[0]):
    blob=re.split(r'[`\-=~!@#$%^&*()_+\[\]{};\\:"|<,./<>? ]',data.iloc[i][0].lower())
    sample=[]
    for j in word_list:
        sample.append(blob.count(j[0]))
        
    features.append(sample)
    if data.iloc[i][1]=='positive':
        labels.append(1)
    else:
        labels.append(0)
```

```python
features=np.array(features)
labels=np.array(labels)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
X_train,X_test,y_train,y_test=train_test_split(features,labels,test_size=0.2,random_state=9)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
print('X_train.shape:',X_train.shape)
print('X_test.shape:',X_test.shape)
print('y_train.shape:',y_train.shape)
print('y_test.shape:',y_test.shape)
```
