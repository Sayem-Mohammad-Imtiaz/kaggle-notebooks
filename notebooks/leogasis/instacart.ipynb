```python
orders_df = pd.read_csv("../input/instacart-market-basket-analysis/orders.csv") #Lectura de base de datos
products_df = pd.read_csv("../input/instacart-market-basket-analysis/products.csv")
aisles_df = pd.read_csv("../input/instacart-market-basket-analysis/aisles.csv")
departmens_df = pd.read_csv("../input/instacart-market-basket-analysis/departments.csv")
order_product_prior_df = pd.read_csv("../input/instacart-market-basket-analysis/order_products__prior.csv")
order_product_train_df = pd.read_csv("../input/instacart-market-basket-analysis/order_products__train.csv")
```

```python
color=sns.color_palette()
plt.figure(figsize=(12,7)) #Especifica las medidas de la figura
sns.countplot(x="order_dow",data=orders_df, color=color[0])
plt.ylabel("Cuenta")
plt.xlabel("Dias de la semana")
plt.title("Frecuencia de orden por día de la semana",fontsize=20)
```

```python
plt.figure(figsize=(14,6))
sns.countplot(x="order_hour_of_day",data=orders_df,color=color[1])
plt.title("Frecuencia de orden por hora",fontsize=14)
plt.ylabel("Cuenta")
plt.xlabel("Hora del día")
plt.show()
```

```python
grouped_df = orders_df.groupby(["order_dow","order_hour_of_day"])["order_number"].aggregate("count").reset_index()
print(grouped_df)
grouped_df = grouped_df.pivot('order_dow',"order_hour_of_day",'order_number')

plt.figure(figsize=(18,7))
sns.heatmap(grouped_df)
```
