## Import packages

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Data

```python
# Get the train data
df_train = pd.read_csv("/kaggle/input/mobile-price-classification/train.csv")
df_train.head()
```

```python
# Training data columns
df_train.columns
```

```python
# Count the values of different classes
df_train['price_range'].value_counts()
```

```python
# Training data info
df_train.info()
```

```python
# Get test data
df_test = pd.read_csv('/kaggle/input/mobile-price-classification/test.csv')
df_test.head()
```

## Splitting data

```python
# Split the training data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df_train.drop('price_range',axis=1), df_train['price_range'],test_size=0.2, random_state=5)
```

## Model building - Logistic regression

```python
# Import packages
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
# Create model and fit it
LR = LogisticRegression()
LR.fit(x_train,y_train)
```

```python
# Predict
y_pred = LR.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

## Model building - KNN

```python
# Import packages
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Create and fit the model
neigh = KNeighborsClassifier(n_neighbors=4)
neigh.fit(x_train, y_train)
```

```python
# Predict
y_pred1 = neigh.predict(x_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
confusion_matrix(y_test,y_pred1)
```

```python
print(classification_report(y_test,y_pred1))
```

## Model building - SVM

```python
# Import packages
from sklearn.svm import SVC
```

```python
# Create and fit the model
svc = SVC(C=1,kernel='linear',gamma = 0.001)
svc.fit(x_train, y_train)
```

```python
# Predict
y_pred2 = svc.predict(x_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
confusion_matrix(y_test,y_pred2)
```

```python
print(classification_report(y_test,y_pred2))
```

## Predictions

Since the SVM model gave the best results, it will be used to predict the test dataset.

```python
# Predict
test_pred = svc.predict(df_test.drop('id',axis=1))
```
