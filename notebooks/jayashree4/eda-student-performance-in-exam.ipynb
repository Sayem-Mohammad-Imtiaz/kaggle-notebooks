## Import packages

```python
# Importing packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data

```python
# Obtaining data
df = pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
df.head()
```

```python
# Find the count of the values
print(df.gender.value_counts())
print(df['race/ethnicity'].value_counts())
print(df['parental level of education'].value_counts())
print(df.lunch.value_counts())
print(df['test preparation course'].value_counts())
```

```python
# Checking for null values
df.isnull().sum()
```

There are no null values.

## Pairplots

Pairplot helps to view the histogram and the correlation scatter plot of the columns.

```python
sns.pairplot(df)
```

As seen from the above plot, there is a linear relation between the reading and writing score.

## Distribution plot

```python
sns.set_style("whitegrid")
sns.displot(df['math score'],kde=True)
```

The math score has an average somewhere near 70.

```python
sns.displot(df['reading score'],kde=True)
```

The reading score has an average somewhere near 80.

```python
sns.displot(df['writing score'],kde=True)
```

The writing score has an average somewhere near 80.

## Catplot

Grouping the students based on the test preparation and gender.

```python
sns.catplot(x="test preparation course",y="math score",hue="gender",data=df)
```

```python
sns.catplot(x="test preparation course",y="reading score",hue="gender",data=df)
```

```python
sns.catplot(x="test preparation course",y="writing score",hue="gender",data=df)
```

```python
sns.catplot(x="test preparation course",y="math score",hue="gender",col="race/ethnicity",data=df)
```

```python
sns.catplot(x="test preparation course",y="reading score",hue="gender",col="race/ethnicity",data=df)
```

```python
sns.catplot(x="test preparation course",y="writing score",hue="gender",col="race/ethnicity",data=df)
```
