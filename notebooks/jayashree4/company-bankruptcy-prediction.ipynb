## Import packages

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Data

```python
df = pd.read_csv('/kaggle/input/company-bankruptcy-prediction/data.csv')
df.head()
```

```python
# Count the value of Bankrupt column
df['Bankrupt?'].value_counts()
```

It is not a balanced dataset since the value of Bankrupt=0 is more than value of Bankrput=1.

```python
# Find if there are any null values
df.isnull().sum()
```

## Model - Logistic regression

```python
# Import packages
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix,classification_report
```

```python
# Create model
LR = LogisticRegression()
```

```python
# Split the dataset
x_train, x_test, y_train, y_test = train_test_split(df.drop('Bankrupt?',axis=1), df['Bankrupt?'],test_size=0.2, random_state=5)
```

```python
# Fit the model
LR.fit(x_train,y_train)
# Predict
y_pred1 = LR.predict(x_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
confusion_matrix(y_test,y_pred1)
```

```python
print(classification_report(y_test,y_pred1))
```

## Model - KNN

```python
# Import packages
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Find the elbow point
test_accuracy = []

for k in range(1,20):
    neigh = KNeighborsClassifier(n_neighbors=k)
    neigh.fit(x_train,y_train)
    test_accuracy.append(accuracy_score(y_test,neigh.predict(x_test)))
```

```python
plt.plot(test_accuracy)
```

```python
# Create model and fit the model
neigh = KNeighborsClassifier(n_neighbors=2)
neigh.fit(x_train, y_train)
```

```python
print(accuracy_score(y_test,neigh.predict(x_test)))
```

```python
print(confusion_matrix(y_test, neigh.predict(x_test)))
```

## Model - Decision tree

```python
# Import packages
from sklearn.tree import DecisionTreeClassifier
```

```python
# Create model and fit
dtree=DecisionTreeClassifier(criterion="gini",max_depth=None,min_samples_leaf=1,random_state=10)
dtree.fit(x_train,y_train)
```

```python
accuracy_score(y_test,dtree.predict(x_test))
```

```python
confusion_matrix(y_test,dtree.predict(x_test))
```

## Model - Random forest

```python
# Import packages
from sklearn.ensemble import RandomForestClassifier
```

```python
# Create model and fit
rfc = RandomForestClassifier(criterion="gini",max_depth=None,min_samples_leaf=1,random_state=10)
rfc.fit(x_train,y_train)
```

```python
accuracy_score(y_test,rfc.predict(x_test))
```

```python
confusion_matrix(y_test,rfc.predict(x_test))
```

## Model - Linear SVM

```python
# Import packages
from sklearn.svm import SVC
```

```python
# Create and fit the model
svc = SVC(C=1,kernel='linear',gamma = 0.001)
svc.fit(x_train,y_train)
```

```python
accuracy_score(y_test,svc.predict(x_test))
```

## Model - Polynomial SVM

```python
# Create and fit the model
svc_p = SVC(C=1,kernel='poly', degree = 2,gamma = 0.001)
svc_p.fit(x_train,y_train)
```

```python
accuracy_score(y_test,svc_p.predict(x_test))
```
