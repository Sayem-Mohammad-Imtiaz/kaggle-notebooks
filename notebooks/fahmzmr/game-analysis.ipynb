## (4) What Genre has been sold the most in a single year?

```python
year_sales = games.groupby(by=['Year', 'Genre']).sum()['Global_Sales'].reset_index()
year_sales2 = year_sales.groupby('Year').transform(max)['Global_Sales'] == year_sales['Global_Sales']
year_sales_max = year_sales[year_sales2].reset_index(drop=True)
year_sales_max.tail(10)
```

```python
genre = year_sales_max['Genre']
```

```python
plt.figure(figsize = (10,6))
graph = sns.barplot(data=year_sales_max[:-3], x='Year', y='Global_Sales')

index = 0
for value in year_sales_max['Global_Sales'][:-3]:
    graph.text(index, value + 1, str(' ' + genre[index] + ' | $' + str(round(value,2))),
    color = '#000', size=10, rotation =90, ha ='center')
    
    index += 1

plt.xticks(rotation=90, size = 12)
plt.xlabel('Year', size = 16)
plt.ylabel('Global Sales (Millions)', size = 16)
plt.show()
```

### Observation: No doubt!! Action game's genre were dominant almost every year and successful than the other game's genre in the market. This insight help us to understand the market and customers needs to secure the new product success and avoid failure.

## (6) Which individual game that have the highest sale price globally?

```python
top_game_sale = games[['Name','Year','Genre','Global_Sales']].sort_values('Global_Sales', ascending = False)[:10]
top_game_sale
```

```python
name = top_game_sale['Name']
year = top_game_sale['Year']
y = np.arange(0,10)
```

```python
plt.figure(figsize=(10, 7))
g = sns.barplot(x='Name', y='Global_Sales', data=top_game_sale)

index=0
for label in top_game_sale['Global_Sales']:
    g.text(index,label-25 ,name[index], rotation ='vertical', size=12, color='#000', ha ='center')
    index += 1

plt.xlabel('Release Year', size = 12)
plt.xticks(y, top_game_sale['Year'], fontsize=12, rotation='vertical')
plt.show()
```

### Observation: 2006 Wii-Sports leading the Global Sales twice then the runne-up Super Mario Bros.

## (7) Overview sales comparison between genre

```python
comp_genre = games[['Genre', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
comp_map = comp_genre.groupby('Genre').sum()
comp_map
```

```python
plt.figure(figsize= (15,10))
sns.heatmap(comp_map, annot = True, fmt= '.1f', linewidths = .5)
plt.xticks(size =14)
plt.yticks(size =14)
plt.xlabel('Sales', size = 14)
plt.ylabel('Genre', size = 14)
plt.show()
```

### Observation: North-America (NA_Sales) have the highest sales compare to the other regions. The insight breaking down the overall market into different market segments based on the characteristic. It helps to choose the right customer for a different product and choosing the most profitable market for the product marketing so we can maximizes sales and profit.

```python
comp_table = comp_map.reset_index()
comp_table = pd.melt(comp_table, id_vars=['Genre'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], var_name='Sale_Area', value_name='Sale_Price')
comp_table.head()
```

```python
plt.figure(figsize=(15, 10))
sns.barplot(x='Genre', y='Sale_Price', hue='Sale_Area', data=comp_table)
plt.xlabel('Genre', size = 14)
plt.ylabel('Sales (Billions)', size = 14)
plt.show()
```

### Observation: Action and Sport sales higher than the others. Also North America have highest sales for every genre. Now we know we the money come from!

## (8) Overview sales comparison between platform

```python
comp_platform = games[['Platform', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].groupby('Platform').sum().reset_index()
comp_platform.head()
```

```python
comp_table = pd.melt(comp_platform, id_vars=['Platform'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], var_name='Sale_Area', value_name='Sale_Price')
comp_table.head()
```

```python
plt.figure(figsize=(30, 15))
sns.barplot(x='Platform', y='Sale_Price', hue='Sale_Area', data=comp_table)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Platform', size = 14)
plt.ylabel('Sales (Billions)', size = 14)
plt.show()
```

### Observation: PS2 and X360 sales higher than the others. And same like the previous chart, North America dominant in sales.

## (10) Top Publisher each year

```python
top_publisher_df = games.groupby(['Year', 'Publisher']).size().reset_index(name='Count')
top_publisher_idx =  top_publisher_df.groupby('Year')['Count'].transform(max) == top_publisher_df['Count']
top_publisher_count = top_publisher_df[top_publisher_idx].reset_index(drop=True)
top_publisher_count  = top_publisher_count.drop_duplicates(subset=["Year", "Count"], keep='last').reset_index(drop=True)
```

```python
publisher= top_publisher_count['Publisher']
```

```python
plt.figure(figsize=(30, 15))
g = sns.barplot(x='Year', y='Count', data=top_publisher_count)
index = 0
for value in top_publisher_count['Count'].values:
    g.text(index, value + 5, str(publisher[index] + ' | ' +str(value)), color='#000', size=14, rotation= 90, ha="center")
    index += 1

plt.xticks(rotation=90)
plt.show()
```

### Observation:  Electronic Arts Inc. was the leader in 2000s and Namco Bandai Games Inc. are conquer the games world in the last decade. 


### Competitive Analysis is very important. It is a strategy where you identify major competitors and research their products, sales and marketing strategy. By doing this company can create solid business strategy to compete and lead the market.
