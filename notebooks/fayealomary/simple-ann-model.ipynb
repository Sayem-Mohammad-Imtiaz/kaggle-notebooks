# **1. Import necessary libraries**

```python
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing 
from sklearn.preprocessing import MinMaxScaler
import seaborn as sns

```

# **2. Import data**

```python
dataframe = pd.read_csv('../input/churn-modelling/Churn_Modelling.csv')
dataframe.head()
```

# **3. Visualizing the data**

```python
df=dataframe.copy()
df.drop(['CustomerId','Surname'],axis=1,inplace=True)
Geo_dummies = pd.get_dummies(prefix='Geo',data=df,columns=['Geography'])
Gender_dummies = Geo_dummies.replace(to_replace={'Gender': {'Female': 1,'Male':0}})
churn_data_encoded = Gender_dummies
sns.countplot(y=churn_data_encoded.Exited ,data=churn_data_encoded)
plt.xlabel("Count of each Target class")
plt.ylabel("Target classes")
plt.show()

churn_data_encoded.hist(figsize=(15,12),bins = 15)
plt.title("Features Distribution")
plt.show()
```

# **4. Preprocessing**

```python
# Transform Geography and Gender to numerical values
le = preprocessing.LabelEncoder()
encoded = dataframe.apply(le.fit_transform)
dataset = encoded.values
# X and Y values
X = dataset[:,3:13]
Y = dataset[:,13]
# Rescale min and max for X
scaler = MinMaxScaler(feature_range=(0, 1))
rescaledX = scaler.fit_transform(X)
```

# **5. Building the Artificial Neural Network**

```python
# initializing the ANN
model = Sequential()
```

```python
#Adding the input layer and the 1st hidden layer
model.add(Dense(12, input_dim=10, kernel_initializer = 'uniform', activation='relu'))
```

```python
#Adding the 2nd hidden layer
model.add(Dense(10, kernel_initializer = 'uniform', activation='relu'))
```

```python
#Adding the output layer 
model.add(Dense(1, kernel_initializer = 'uniform', activation='relu'))
```

```python
#Compiling the ANN 
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train model
history = model.fit(rescaledX, Y, epochs=50, batch_size=50,  verbose=1)
```

```python
# Print Accuracy
scores = model.evaluate(rescaledX, Y) 
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```
