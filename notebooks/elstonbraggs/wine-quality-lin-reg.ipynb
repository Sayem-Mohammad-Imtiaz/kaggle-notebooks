```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

wine= "../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv"
df= pd.read_csv(wine)
```

```python

df.head()
df.describe()
df.info()
```

```python

sns.histplot(df['quality'])
```

```python
sns.pairplot(df,hue='quality',palette="coolwarm")
```

```python
X = df[['alcohol']]
y = df['quality']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

lin_reg = LinearRegression()

lin_reg.fit(X_train,y_train)
```

```python

print(f'Coeff {lin_reg.coef_}')
print(f'Intercept {lin_reg.intercept_}')
```

```python
y_pred = lin_reg.predict(X_test)
```

```python
plt.scatter(X_test,y_test, color = 'blue')
plt.scatter(X_test,y_pred, color= 'red')
plt.xlabel('alcohol')
plt.ylabel('quality')
plt.title('alcohol Vs wine_quality')
plt.show()
```

```python
# confirm the model
sns.distplot((y_pred - y_test))

my_alcohol = 1
lin_reg.predict(np.array(my_alcohol).reshape(-1,1))
```
