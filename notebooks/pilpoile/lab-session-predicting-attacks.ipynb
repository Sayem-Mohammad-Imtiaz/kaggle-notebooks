# **6 Which columns are categories? List them; extract existing values.**

```python
print(df.dtypes[df.dtypes == 'object'].index)
for category in df.dtypes[df.dtypes == 'object'].index:
    print(category)
    print(list(set(df[category])))
```

# **7 Which columns are numeric? List them; extract min, max, mean, median and standard deviation values for ‘rate’.**

```python
newdf = df._get_numeric_data()
print(newdf.columns)
for category in newdf.columns:
    print('>> '+category)
    print('min: '+str(min(df[category])))
    print('max: '+str(max(df[category])))
    print('mean: '+str(np.median(df[category])))
    print('std: '+str(np.std(df[category])))
```

*newdf is create for work only with metric data by using _get_numeric_data. All object are remove in newdf*
