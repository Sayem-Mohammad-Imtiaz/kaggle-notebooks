```python
column_names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'PRICE']
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
data = pd.DataFrame(boston.data)
```

```python
data.columns = boston.feature_names
data.head()
```

```python
data['PRICE'] = boston.target 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
sns.boxplot(data=data['PRICE'])
```

```python
data = data[~(data['PRICE'] >= 40.0)] # removed the outliers
print(np.shape(data))
```

```python
fig, ax = plt.subplots(nrows =2, ncols= 7, figsize =(16,4))

col = data.columns
index = 0

for i in range(2):
    for j in range(7):
        sns.distplot(data[col[index]], ax = ax[i][j])
        index= index +1
plt.tight_layout()
```

```python
corrmat = data.corr()
corrmat
```

```python
fig, ax = plt.subplots(figsize=(18,10))
sns.heatmap(corrmat.abs(), annot = True)
```

```python
#l =['CRIM','ZN','CHAS','AGE','DIS','RAD','B']
l =['CHAS','DIS']

for i in l:
    data.drop(i, axis =1, inplace = True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop('PRICE', axis =1)
y = data['PRICE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.3, random_state =100)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,y_predict)
score
```

```python
import xgboost as xgb
```

```python
xgb = xgb.XGBRegressor(n_estimators=100, learning_rate=0.08, gamma=0, subsample=0.75,
                           colsample_bytree=1, max_depth=7)
```

```python
xgb.fit(X_train,y_train)
```

```python
predictions = xgb.predict(X_test)
score = r2_score(y_test,predictions)
```
