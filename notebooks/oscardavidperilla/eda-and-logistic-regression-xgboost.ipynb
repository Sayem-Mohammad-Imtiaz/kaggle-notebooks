```python
df_copy = df
```

```python
df_copy.replace(np.nan, np.mean(df_copy), inplace = True)
```

```python
ax = plt.figure(figsize=(32,15))
corr = df_copy.corr()
ax = sns.heatmap(corr, annot= True)
```

```python
x = df_copy.iloc[:, 0:-1].values
y = df_copy.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split as tts
x_train, x_test, y_train, y_test = tts(x, y, test_size = 0.25, random_state = 2)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x_train, x_test = sc_x.fit_transform(x_train), sc_x.transform(x_test)
```

```python
from xgboost import XGBClassifier
classifier = XGBClassifier(learning_rate = 0.001).fit(x_train, y_train)
y_pred1 = classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
ac1 =accuracy_score(y_test, y_pred1)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression().fit(x_train, y_train)
y_pred = classifier.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.svm import SVC
classifier = SVC().fit(x_train, y_train)
y_pred = classifier.predict(x_test)
```

```python
ac = accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
ax = plt.figure(figsize = (32,15))
sns.heatmap(cm, annot =True)
```
