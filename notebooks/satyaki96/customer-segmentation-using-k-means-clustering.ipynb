```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('../input/customer-segmentation-tutorial-in-python/Mall_Customers.csv')
dataset.head()
```

**K-Means Clustering**

```python
X=dataset.iloc[:,3:]
X.head()
```

```python
X.iloc[:,0].plot(kind = 'hist')
plt.xlabel('Annual Income')
plt.show()
```

```python
X.iloc[:,1].plot(kind = 'hist')
plt.xlabel('Spending score')
plt.show()
```

To Find the Optimal Cluster we will use elbow method

```python
#Elbow method
from sklearn.cluster import KMeans
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++', random_state = 75)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The elbow method')
plt.xlabel('The number of Clusters')
plt.ylabel('Wcss')
plt.show()
```

```python
#Here the optimal value of k is 5.
kmeans=KMeans(n_clusters=5,init='k-means++', random_state = 96)
y_kmeans= kmeans.fit_predict(X)
print(y_kmeans)
```

```python
#Visualising the clusters
plt.scatter(X[y_kmeans==0]['Annual Income (k$)'], X[y_kmeans==0]['Spending Score (1-100)'], s = 100,c = 'red', label = 'Cluster 1')
plt.scatter(X[y_kmeans==1]['Annual Income (k$)'], X[y_kmeans==1]['Spending Score (1-100)'], s = 100,c = 'blue', label = 'Cluster 2')
plt.scatter(X[y_kmeans==2]['Annual Income (k$)'], X[y_kmeans==2]['Spending Score (1-100)'], s = 100,c = 'green', label = 'Cluster 3')
plt.scatter(X[y_kmeans==3]['Annual Income (k$)'], X[y_kmeans==3]['Spending Score (1-100)'], s = 100,c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_kmeans==4]['Annual Income (k$)'], X[y_kmeans==4]['Spending Score (1-100)'], s = 100,c = 'black',label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], s = 300, c = 'yellow', label = 'centroids')
plt.title('Customer Clusters')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```

```python
from sklearn import metrics
print(metrics.silhouette_score(X,y_kmeans))
```
