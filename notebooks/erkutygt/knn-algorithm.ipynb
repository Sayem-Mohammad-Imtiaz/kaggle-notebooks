```python
#import data
data = pd.read_csv("/kaggle/input/biomechanical-features-of-orthopedic-patients/column_3C_weka.csv")
```

```python
data
```

```python
data.info()
```

```python
y = data["class"].values
y
```

```python
data["class"] = [0 if each == "Normal" else 1 for each in data["class"]]
```

```python
x_data = data.drop(["class"], axis = 1)
x_data
```

# Normalization


```python
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))
x
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=1)
```

# KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=13)
knn.fit(x_train, y_train)
prediction = knn.predict(x_test)
```

```python
prediction
```

```python
print("{} nn score: {}".format(3, knn.score(x_test,y_test)*100))
```

```python
# find k value
score_list = []

for each in range(1,15):
    knn2 = KNeighborsClassifier(n_neighbors=each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))
    
plt.plot(range(1,15),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```
