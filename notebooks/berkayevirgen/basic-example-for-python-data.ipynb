```python
# local and global scope subject
x = 5
def f():
    x = 4
    return x
print(x) # this will return 5 > because this is global scope
print(f()) # this will return 4 > because this is local scope
```

```python
# We should just some look nested Func (func inside func)
def square():
    def add():
        x,y = 3,7
        z = x+y
        return z
    return add()**2
print(square())
```

```python
# DEFAULT and FLEXIBLE ARGUMENTS
def f(a,b=1,c=2):
    y = a+b+c
    return y
print(f(7))
# can we change default args?
print(f(7,5,5)) # so, yes
```

```python
# flexible args *args
def f(*args):
    for i in args:
        print(i)
f(55)
print('\n')
f(1,2,3,4)

# flexible args **kwargs > dictionary
def f(**kwargs):
    for key, value in kwargs.items():
        print(key, " ", value)
f(country = 'Turkey', capital='Ankara', population='Over 5 million')
```

```python
# shoul not forget nested function (func inside func)
def square():
    def add():
        x = 5
        y = 7
        z = x + y
        return z
    return add()**2
print(square()) 
```

```python
# Arguments
# default arguments
def f(a, b = 4, c = 6):
    y = a + b + c
    return y
print(f(5))
# can we change default arguments
print(f(5,8,5)) # so, yes
```

```python
# flexible arguments *args
def f(*args):
    for i in args:
        print(i)
f(1)
print("\n")
f(1,2,3,4,5,6)

# flexible arguments **kwargs >> dictionary

def f(**kwargs):
    for key, value in kwargs.items():
        print(key, " ", value)
f(country = 'Turkey', capital = 'Ankara', population = 'Over 6m')
```

```python
# zip example
l1 = [2,3,4,5]
l2 = [6,7,8,9]
z = zip(l1,l2)
print(z)
z_list = list(z)
print(z_list)
```

```python
# go unzip
uz=zip(*z_list)
ul1,ul2 = list(uz) # unzip returns tuple
print(ul1)
print(ul2)
print(type(ul2))
```

```python
# list comprehension
# this is important topic
num1 = [1,2,3]
num2 = [i + 1 for i in num1 ]
print(num2)
```

```python
# conditionals on iterable
num1 = [3,4,5]
num2 = [i**2 if i == 4 else i-3 if i < 4 else i*4 for i in num1]
print(num2)
```
