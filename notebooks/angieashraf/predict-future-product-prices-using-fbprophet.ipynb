# # **Data Preparation for using FaceBook prophet**

```python
avocado_df
```

```python
avocado_prophet_df=avocado_df[['Date','AveragePrice']].rename(columns={'Date':'ds','AveragePrice':'y'})
```

```python
avocado_prophet_df
```

```python
m=prophet.Prophet()
m.fit(avocado_prophet_df)
```

```python
#forecast into the future
future=m.make_future_dataframe(periods=30)
forecast=m.predict(future)
```

```python
forecast
```

```python
#visualize our data
figure=m.plot(forecast,xlabel='Date',ylabel='Price')
```

```python
#the visualization of performance monthly in a year
plt.figure(figsize=(30,10))
figure2=m.plot_components(forecast)
```

# # Applying forecast on specific region

```python
avocado_df_sample=avocado_df[avocado_df['region']=='West']
avocado_df_sample
```

```python
avocado_df_sample=avocado_df_sample.sort_values('Date')
```

```python
avocado_df_sample=avocado_df_sample[['Date','AveragePrice']].rename(columns={'Date':'ds','AveragePrice':'y'})
```

```python
m2=prophet.Prophet()
m2.fit(avocado_df_sample)
```

```python
future2=m2.make_future_dataframe(periods=365)
forecast=m2.predict(future2)
```

```python
forecast
```

```python
figure=m2.plot(forecast,xlabel='Date',ylabel='Price')
```

```python
figure3=m2.plot_components(forecast)
```
