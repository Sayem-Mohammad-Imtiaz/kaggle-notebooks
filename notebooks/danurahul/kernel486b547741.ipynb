```python
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow import feature_column
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split
```

```python
dataframe = pd.read_csv('/kaggle/input/covid19-corona-virus-india-dataset/complete.csv')
```

```python
train, test = train_test_split(dataframe, test_size=0.20)
train, val = train_test_split(train, test_size=0.20)
```

```python
print(len(train))
print(len(test))
print(len(val))
```

**create INPUT PIPELIINE WITH tf.data**

```python
def df_to_dataset(dataframe, shuffle=True, batch_size=32):
    dataframe=dataframe.copy()
    labels= dataframe.pop('New recovered')
    ds = tf.data.Dataset.from_tensor_slices((dict(dataframe), labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(dataframe))
    ds = ds.batch(batch_size)
    return ds
```

```python
train_ds=df_to_dataset(train)
val_ds=df_to_dataset(val)
test_ds=df_to_dataset(test)
```
