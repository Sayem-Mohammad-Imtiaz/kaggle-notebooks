# IMPORTING THE DATA

```python
import pandas as pd
data=pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')
data.head(2)
```

# DATA PREPROCESSING

```python
data.isnull().sum()
```

```python
data=data.fillna('unknown')
data.isnull().sum()
```

Splitting the unique casts in each and every rows of the cast column in the data

```python
cast_series=data['cast'].str.split(',')
cast_series[:3]
```

Summing up all the rows contains list in above 'cast_series'.

```python
casts=[]
for i in cast_series:
    casts=casts+i
len(casts),casts[:10]    
```

```python
data.columns
```

Dropping the cast column in column names and creating the empty dictionary with the column keys inside them

```python
col=data.columns.drop('cast')
dic=dict()
for i in col:
    dic['{}'.format(i)]=[]
dic    
```

Creating the duplicates data for the each and every caste in non unique 'casts_series' casts. 

```python

for i,k in zip(cast_series,range(len(cast_series))):
    for l in col:
        dic['{}'.format(l)]= dic['{}'.format(l)]+[data[l][k] for j in range(len(i))]    
dic['cast']=casts 
df_unique_cast=pd.DataFrame(dic)
df_unique_cast.head(2)
```

# AS SAME AS CASTS FOR COUNTRY

```python
country_series=data['country'].str.split(',')
country_series[:3]
country=[]
for i in country_series:
    country=country+i
len(country),country[:2]    
```

```python
col=data.columns.drop('country')
dic=dict()
for i in col:
    dic['{}'.format(i)]=[]
for i,k in zip(country_series,range(len(country_series))):
    for l in col:
        dic['{}'.format(l)]= dic['{}'.format(l)]+[data[l][k] for j in range(len(i))]    
dic['country']=country 
df_unique_country=pd.DataFrame(dic)
df_unique_country.head(2)
```

# FOR THE DIRECTORS

```python
director_series=data['director'].str.split(',')
director=[]
for i in director_series:
    director=director+i 
col=data.columns.drop('director')
dic=dict()
for i in col:
    dic['{}'.format(i)]=[]
for i,k in zip(director_series,range(len(director_series))):
    for l in col:
        dic['{}'.format(l)]= dic['{}'.format(l)]+[data[l][k] for j in range(len(i))]    
dic['director']=director 
df_unique_director=pd.DataFrame(dic)
df_unique_director.head(2)
```

# FOR THE GENRE

```python
genre_series=data['listed_in'].str.split(',')
genre=[]
for i in genre_series:
    genre=genre+i 
col=data.columns.drop('listed_in')
dic=dict()
for i in col:
    dic['{}'.format(i)]=[]
for i,k in zip(genre_series,range(len(genre_series))):
    for l in col:
        dic['{}'.format(l)]= dic['{}'.format(l)]+[data[l][k] for j in range(len(i))]    
dic['genre']=genre 
df_unique_genre=pd.DataFrame(dic)
df_unique_genre.head(2)
```

# VISUALIZATIONS OF INSIGHTS

```python
y=list(set(df_unique_cast['cast']))
df_cast=pd.DataFrame({'cast':y,'No of movies/Tv shows casted':[casts.count(i) for i in y]})
```

```python
df_cast=df_cast.sort_values('No of movies/Tv shows casted')
```

```python
#I am ignoring unknow cast cause there is no point when we choose unknown cast as most movies/tvshow casted
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
x=[i+1 for i in range(10)]
plt.bar(x,df_cast['No of movies/Tv shows casted'][-11:-1],color=['r','b','g','y','orange','black','brown','y','r','b'])
plt.xticks(x,df_cast['cast'][-11:-1],rotation=90)
plt.xlabel('Cast')
plt.ylabel('No of movies/Tv shows casted')
plt.title('Top 10 Cast')
```

So we conclude that Anupam kher casted the most no of movies or tv shows 

```python
y=list(set(df_unique_country['country']))
df_country=pd.DataFrame({'country':y,'No of movies/Tv shows in country':[country.count(i) for i in y]})
df_country=df_country.sort_values('No of movies/Tv shows in country')
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
x=[i+1 for i in range(10)]
plt.bar(x,df_country['No of movies/Tv shows in country'][-10:],color=['r','b','g','y','orange','black','brown','y','r','b'])
plt.xticks(x,df_country['country'][-10:],rotation=90)
plt.xlabel('Country')
plt.ylabel('No of movies/Tv shows in country')
plt.title('Top 10 Countries')
```

So we coclude that most of the tv shows and movies released in UNITED STATES

```python
y=list(set(df_unique_director['director']))
df_director=pd.DataFrame({'director':y,'No of movies/Tv shows directed':[director.count(i) for i in y]})
df_director=df_director.sort_values('No of movies/Tv shows directed')
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
x=[i+1 for i in range(10)]
#THERE ARE SO MANY UNKOWN DIRECTORS SO I REMOVE UNKNOWN
plt.bar(x,df_director['No of movies/Tv shows directed'][-11:-1],color=['r','b','g','y','orange','black','brown','y','r','b'])
plt.xticks(x,df_director['director'][-11:-1],rotation=90)
plt.xlabel('Director')
plt.ylabel('No of movies/Tv shows Directed')
plt.title('Top 10 Directors')
```

So we conclude that Jan Suter Direvted the most Tvshows/Movies
