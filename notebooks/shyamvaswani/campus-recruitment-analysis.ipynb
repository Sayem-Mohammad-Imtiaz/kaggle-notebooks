# Instantiate Classifiers

```python
#Classifiers List
clf_names = []
gboost =GradientBoostingClassifier(random_state=77) 
rf = RandomForestClassifier(random_state=77) 
svm_clf = SVC(random_state=77) # classification SVC
logr = LogisticRegression(max_iter=2000,random_state=77)
#lr= LinearRegression(random_state=77) # regression only
knn = KNeighborsClassifier()
nn=MLPClassifier(alpha=1,random_state=77,max_iter=400000) # Neural Networks
ada =AdaBoostClassifier(random_state=77)
dt= DecisionTreeClassifier(random_state=77)
gpc=GaussianProcessClassifier(1.0 * RBF(1.0),random_state=77)
svc=SVC(kernel="linear", C=0.025,probability=True,random_state=77) # Linear SVM
nb= GaussianNB()
qda= QuadraticDiscriminantAnalysis()
svm_rbf= SVC(kernel='rbf',gamma=2, C=1,probability=True,random_state=77) #RBF SVM
sgd=SGDClassifier(max_iter=1000, tol=10e-3,penalty='elasticnet',random_state=77)

```

# Fit, Predict,Score

```python
# Pipeline

def fit_predict_score(model, X_y_train_test):
    X_train, X_test, y_train, y_test = X_y_train_test
    # extracts and flattens the series y_train which has series and values to values only
    model.fit(X_train, y_train.values.ravel())
    y_pred = model.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    print('Accuracy Score on test data :',score)
    return 
```

# Normalization (MinMaxScaler)

```python
# Pipeline with MinMaxScaler()

def MMScale_fit_predict_score(model, X_y_train_test):
    X_train, X_test, y_train, y_test = X_y_train_test
    scaler = MinMaxScaler() 
    X_train_scaled = scaler.fit_transform(X_train) 
    # extracts and flattens the series y_train which has series and values to values only
    model.fit(X_train_scaled, y_train.values.ravel()) 
    X_test_scaled = scaler.transform(X_test) 
    y_pred = model.predict(X_test_scaled) 
    score = accuracy_score(y_test, y_pred)
    print('Accuracy Score on test data :',score)
    return 
```
