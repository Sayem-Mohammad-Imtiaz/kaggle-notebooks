```python
hemanth=pd.read_csv("/kaggle/input/titanic/train_data.csv")
hemanth.head()
```

```python
y=hemanth.filter(['Survived'])
y.head()
hemanth.drop(['Survived'],axis=1,inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.heatmap(hemanth.corr())
```

```python
def correlation(data,thresold):
    data=data.corr()
    corr_features= set()
    for i in range(len(data.columns)):
        for j in range(i):
            if(abs(data.iloc[i,j])>thresold):
                colname=data.columns[i]
                corr_features.add(colname) 
    return corr_features
```

```python
features=correlation(hemanth,0.7)
print(features)
```

```python
hemanth.drop(features,axis=1,inplace=True)
```

```python
hemanth.head()
```

```python
hemanth.drop(['PassengerId'],axis=1,inplace=True)
```

```python
hemanth.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
X=hemanth
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
X_train.head()
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```
