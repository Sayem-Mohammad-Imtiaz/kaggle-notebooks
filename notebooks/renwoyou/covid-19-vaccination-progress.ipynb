```python
import pandas as pd
import plotly.express as px
```

```python
data_df = pd.read_csv("/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv")
```

```python
country_vaccine = data_df.groupby(["country", "iso_code", "vaccines"])['total_vaccinations',                          
'total_vaccinations_per_hundred',                                                       
'daily_vaccinations',
'daily_vaccinations_per_million',
'people_vaccinated',
'people_vaccinated_per_hundred'].max().reset_index()
country_vaccine.columns = ["Country", "iso_code", "Vaccines", "Total vaccinations", "Percent",
"Daily vaccinations", "Daily vaccinations per million", "People vaccinated", "People vaccinated per hundred"]
```

# What vaccines are used and in which countries?

```python
fig = px.choropleth(locations=country_vaccine['Country'], 
                    locationmode="country names",
                    color=country_vaccine['Vaccines'],
                    title="Countries using each vaccine (different colors for each vaccine)",
                   )
fig.update_layout({'legend_orientation':'h'})
fig.update_layout({'legend_title':'Vaccine scheme'})
fig.show()
```

**Vaccination total of the country**

You can put different data in the code.
Therefore, I do not show all of them.

```python
country_vaccine = country_vaccine.sort_values("Total vaccinations",ascending=False)
fig = px.bar(country_vaccine,x = "Vaccines", y = "Total vaccinations", color = "Country")
fig.show()
```

```python
fig = px.treemap(country_vaccine,path =["Vaccines","Country"] , values = "Total vaccinations")
fig.show()
```

Vaccination total per country

```python
fig = px.bar(country_vaccine,x = "Country", y = "Total vaccinations",title ="Total vaccinations")
fig.update_xaxes(tickangle=45)
fig.show()
```

People vaccinated per hundred per country

```python
country_vaccine = country_vaccine.sort_values("People vaccinated per hundred",ascending=False)
fig = px.bar(country_vaccine,x = "Country", y = "People vaccinated per hundred",title = "People vaccinated per hundred")
fig.update_xaxes(tickangle=45)
fig.show()
```
