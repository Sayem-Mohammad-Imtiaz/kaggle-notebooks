```python
import pandas as pd
```

```python
data = pd.read_csv("../input/water-potability/water_potability.csv")
```

```python
data.head()
```

```python
data.isnull().sum()  # hangi sütunda ne kadar boş veri olduğunu gösterir.
```

## ph, sülfat ve trihalometan değerlerinin ortalamısını alıp null olan değerlere ortalama değerle dolduruyoruz.

```python
mean_ph = data["ph"].mean()  # ph ortalaması

data["ph"] = data["ph"].fillna(mean_ph)
```

```python
mean_sulfate = data["Sulfate"].mean()

data["Sulfate"] = data["Sulfate"].fillna(mean_sulfate)
```

```python
mean_Trihalomethanes = data["Trihalomethanes"].mean()

data["Trihalomethanes"] = data["Trihalomethanes"].fillna(mean_Trihalomethanes)
```

```python
ytrain = data.pop("Potability")
xtrain = data
```

```python
xtrain.columns  # kaç parametre olduğunuz söyler.
```

# Data Split

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(xtrain, ytrain, test_size = 0.2)
```

# Training

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=3)
rf.fit(x_train,y_train)
```

# Testing

```python
from sklearn.metrics import accuracy_score

y_test_pred = rf.predict(x_test)
accuracy_score(y_test, y_test_pred)
```

# Saving

```python
import pickle

rf_model_pkl = open("./water_quality_prediction.pkl", 'wb')
pickle.dump(rf, rf_model_pkl)
rf_model_pkl.close()
```
