```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_csv('../input/netflix-shows/netflix_titles.csv')
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
colors = ['red', 'yellow']
displace = [0, 0.1]
df.type.value_counts(sort=False).plot.pie(autopct='%1.1f%%', colors=colors, explode=displace, figsize=(20,10), textprops={'fontsize': 18}, labels=None)
plt.legend(labels=df.type.unique(), fontsize=20)
plt.axis('equal')
plt.title('Percentage of Movies and TV Shows in Netflix', fontsize=20)
plt.show()
```

<h3>In 2019 Netflix has 68.4% movies and TV shows has 31.6%</h3>

```python
dir_list = pd.Series(np.random.choice(list('director'), 1000))
director = df.director.value_counts(ascending=False)
colors = ['r','b','b','b','b','b','b', 'b', 'b', 'b']
director[:10].plot.barh(color=colors, align='center', figsize=(10,10)).invert_yaxis()
plt.title('Top 10 directors with most directed shows in Netflix')
plt.xlabel('No. of Product')
```

<H3> Raul Campos and Jan Suter has the most number of directed shows in Netflix

```python
country_list = pd.Series(np.random.choice(list('director'), 1000))
country = df.country.value_counts(ascending=False)
country[:5].plot.pie( autopct='%1.1f%%', figsize=(24,12), textprops={'fontsize': 18})
plt.legend(labels=country[:5], title='Produced by numbers')
plt.title('Top 5 countries with the largest number of TV and Movie Productions', fontsize=18)
centre_circle = plt.Circle((0,0),0.35,color='black', fc='white',linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
```

<h3>United States is the leading country who produce TV and Movie Production. </h3>

```python
rating_list = pd.Series(np.random.choice(list('rating'), 1000))
rating = df.rating.value_counts(ascending=False)
colors = ['r','b','b','b','b','b','b', 'b', 'b', 'b','b','b','b','b','b']
rating.plot.barh(color=colors, figsize=(10,10)).invert_yaxis()
plt.title('Number of Shows with their Ratings')
plt.xlabel('No. of Product')
```

<h3>TV-MA (Parental Guidance) is the most common ratings of all netflix shows</h3>

```python
sns.distplot(df.loc[df['release_year'] > 2000, 'release_year'], fit=norm, kde=False)
plt.title('Number of Movies and TV shows produced every 5 years')
```
