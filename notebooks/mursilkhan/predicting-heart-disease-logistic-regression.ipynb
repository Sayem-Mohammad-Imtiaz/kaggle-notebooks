# Data Preprocessing

Taking care of missing data by replacing missing values with
* Education: Highest Occurence
* CigsPerDay: Highest Occurence
* TotChol: Mean
* BMI: Mean
* BPMeds: Highest Occurence
* Glucose: Mean

```python
#Storing the data as an array
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
print(y)
```

```python
from sklearn.impute import SimpleImputer
#using mean to replace missing values
mean_imputer = SimpleImputer(missing_values=np.nan, strategy ='mean')
mean_imputer.fit(X[:,[9,12,14]])
X[:,[9,12,14]]= mean_imputer.transform(X[:,[9,12,14]])

#using most_frequent value to replace missing data
#most_frequent
freq_imputer = SimpleImputer(missing_values=np.nan, strategy ='most_frequent')
freq_imputer.fit(X[:, [2,4,5]])
X[:,[2,4,5]]= freq_imputer.transform(X[:,[2,4,5]])
```

Splitting the dataset into test and train set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state= 0)
```
