```python
Abnormal = data[data["class"] == "Abnormal"]
Normal = data[data["class"] == "Normal"]
```

```python
plt.scatter(Abnormal.pelvic_incidence, Abnormal.pelvic_radius, color = "red", label = "Abnormal", alpha = 0.3)
plt.scatter(Normal.pelvic_incidence, Normal.pelvic_radius, color = "blue", label = "Normal", alpha = 0.3)
plt.xlabel("Pelvic Incidence")
plt.ylabel("Pelvic Radius")
plt.legend()
plt.show()
```

```python
# Assign the Abnormal values as 1, Normal values as 0 
data["class"] = [1 if each == "Abnormal" else 0 for each in data["class"]]
```

```python
y = data["class"].values
x_data = data.drop(["class"], axis=1)
```

****Normalization****
* (x - min(x))-(max(x) - min(x))

```python
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))
```

****TRAIN-TEST SPLIT****

```python
from sklearn.model_selection import train_test_split

# Split 80% of the data to train, 20% of the data to test -->  test_size = 0.2 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

*Firstly, we should find the right k number that gives us a good accuracy.*
* I first looked at the graph which has range(1,100) and the highest accuracy seemed to be between 1 and 30, So I made it range(1,30)

```python
from sklearn.neighbors import KNeighborsClassifier

# Finding k value
score_list = []
for each in range(1,30):
    knn1 = KNeighborsClassifier(n_neighbors = each)
    knn1.fit(x_train, y_train)
    score_list.append(knn1.score(x_test, y_test))

```

```python

plt.plot(range(1,30), score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```

**As you can see, k value that has the best accuracy is 15.**

```python
knn = KNeighborsClassifier(n_neighbors = 15)
knn.fit(x_train, y_train)
prediction = knn.predict(x_test)

print("{} knn score {}".format(15, knn.score(x_test, y_test)))
```
