**"BMI": Body Mass Index**<a id='10'></a>

Body mass index (BMI) is a measure of body fat based on height and weight.


Now we visualize some more dependencies with the point of view of "bmi".

```python
#Distribution of charges for patients with BMI greater than 30
sns.distplot(df[(df.bmi >= 30)]['charges'], # kde=True,,cumulative=True
             color = 'm')
             
```

```python
#Distribution of charges for patients with BMI less than 30
sns.distplot(df[(df.bmi < 30)]['charges'], color = 'b')
```

Patients with BMI above 30 spend more on treatment!

```python
g = sns.jointplot(x="bmi", y="charges", #kind="kde",
                  data = df, color="r")
```

```python
ax = sns.scatterplot(x='bmi',y='charges',data=df,palette='magma',hue='smoker')
```

```python
sns.lmplot(x="bmi", y="charges", hue="smoker", data=df, palette = 'rainbow')
```

So we can say that bmi has a linear effect on the charges for smokers and non-smokers separately.

**CATEGORIZING BMIs**<a id='11'></a>

BMI CATEGORIES

* Underweight = <18.5
* Normal weight = 18.5â€“25
* Overweight = 25â€“30
* Obesity = BMI of 30 or greater.


```python
data = df.copy(deep=True)
```

```python
bmi_cat = []
for i in df.bmi:
    if i<=18.5:
        bmi_cat.append("Underweight")
    elif (i> 18.5 and i<=25):
        bmi_cat.append("Normal weight")
    elif (i>25 and i<30):
        bmi_cat.append("Overweight")
    else:
        bmi_cat.append("Obesity")
data["BMI_Category"]=bmi_cat
data.sample(10)
```

```python
sns.catplot(x="BMI_Category", y="charges", hue="sex",
             data=data)
```

Clearly, charges are high for patients with obesity.
