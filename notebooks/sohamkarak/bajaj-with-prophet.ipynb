```python
df=pd.read_csv('../input/nifty50-stock-market-data/BAJAJFINSV.csv')
```

```python
df.head()
```

```python
df=df[['Date','VWAP']]
```

```python
df.columns
```

```python
df.columns=['ds','y']
```

```python
df['ds']=pd.to_datetime(df.ds,format='%Y-%m')
```

```python
plt.figure(figsize = (16, 7))
plt.plot(df.ds,df.y)
plt.show()
```

```python
m=Prophet(interval_width=0.95)
m.fit(df)
```

```python
future=m.make_future_dataframe(periods=50, freq='MS')
```

```python
forecast=m.predict(future)
```

```python
preds= forecast['yhat'][:-50]
```

```python
plt.figure(figsize=(15,6))
plt.plot(df.ds,preds,color='red',label=' Predictions')
plt.plot(df.ds,df.y,color='blue',label='Actual')
plt.legend()
plt.show()
```
