I wanted to compare the values of Health, Trust  and Economy

I added a new feature as COMPRISE. I wanted to do this in a new feature; I take avarege economy value from all dataset and then I compared it with all the values in this dataset,
If my value is higher than economy value and then COMPRISE column is HIGH else LOW.

```python
AVG_ECONOMY = data.Economy.mean()
data["COMPRISE"]=["LOW" if AVG_ECONOMY> each else "HIGH" for each in data.Economy]
a=data["COMPRISE"]=="HIGH"
data[a]
```
