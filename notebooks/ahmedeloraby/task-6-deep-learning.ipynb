Removing units from engine, mileage and max_power

```python
pd.get_dummies(dataSet['engine']).sum()
```

```python
dataSet['engine'] = pd.to_numeric(dataSet['engine'].apply(lambda x: str(x)[:-3]), errors='coerce').astype(np.int64)
dataSet
```

```python
balabizo = pd.to_numeric(dataSet['max_power'].apply(lambda x: str(x)[:-4]), errors='coerce')
balabizo.isnull().sum()
dataSet['max_power'] = balabizo
dataSet
```

```python
balabizo = pd.to_numeric(dataSet['mileage'].apply(lambda x: str(x)[:-5]), errors='coerce')
balabizo.isnull().sum()
dataSet['mileage'] = balabizo
dataSet
```

```python
dataSet.info()
```

```python
dataSet.nunique()
```

Quantifying torque, owner ,fuel and seller_type

```python
dataSet['torque'] = dataSet['torque'].factorize()[0]
dataSet['owner'] = dataSet['owner'].factorize()[0]
dataSet['fuel'] = dataSet['fuel'].factorize()[0]
dataSet['seller_type'] = dataSet['seller_type'].factorize()[0]
dataSet['name'] = dataSet['name'].factorize()[0]
dataSet
```

```python
dataSet.info()
```

splitting transmission into two columns

```python
transmdf=pd.get_dummies(dataSet['transmission'])

dataSet[transmdf.columns]=transmdf
dataSet
```

```python
dataSet  =dataSet.drop(['transmission'],axis = 1)
```

```python
dataSet  =dataSet.drop(['name'],axis = 1)
```

Splitting Data

```python
X_train, X_test, y_train, y_test = train_test_split( dataSet.drop('selling_price', axis=1), dataSet['selling_price'], test_size=0.2, random_state=42)
```

Normalization:

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
X_train= sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
X_train.shape
```

```python
model = keras.Sequential(
    [
        Dense(16, name="input", input_shape=(12,), activation = 'relu' ),
        Dense(32, name="hidden1", input_shape=(12,), activation = 'relu' ),
        Dense(1, name="output", activation = 'linear')
    ]
)
model.summary()
```

```python
model.compile(loss='mean_squared_logarithmic_error',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=100)
```

```python
model.evaluate(X_test, y_test)
```
