```python
df['Reason']=df['title'].apply(lambda title: title.split(":")[0])
df.head()
```

What is the most common Reason for a 911 call based off of this new column?

```python
df['Reason'].value_counts()
```

using seaborn to create a countplot of 911 calls by Reason. 


```python
plt.subplots(figsize=(12,6))
sns.countplot(x='Reason',data=df)
```

 Using pd.to_datetime to convert the column from strings to DateTime objects.

```python
df['timeStamp']=pd.to_datetime(df['timeStamp'])
```


 Now that the timestamp column are actually DateTime objects, use .apply() to create 3 new columns called Hour, Month, and Day of Week.  Creating these columns based off of the timeStamp column
    
    

```python
time = df['timeStamp'].iloc[0]
df['hour']=df['timeStamp'].apply(lambda time:time.hour)

df['month']=df['timeStamp'].apply(lambda time: time.month)

df['day of week']=df['timeStamp'].apply(lambda time: time.dayofweek)
```

Using the .map() with this dictionary to map the actual string names to the day of the week:

```python
df['day of week']=df['day of week'].map( {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'})
```

 using seaborn to create a countplot of the Day of Week column with the hue based off of the Reason column.

```python
plt.subplots(figsize=(12,6))
sns.countplot(x='day of week',data=df,hue='Reason')
```

Creating a new column called 'Date' that contains the date from the timeStamp column.

```python
df['date']=df['timeStamp'].apply(lambda time: time.date())
df.head()
```

 groupby this Date column with the count() aggregate and creating a plot of counts of 911 calls

```python
Date=df.groupby('date').count()
Date.head()
plt.subplots(figsize=(12,6))
Date['hour'].plot()
```

 creating heatmaps with seaborn and our data. We'll first need to restructure the dataframe so that the columns become the Hours and the Index becomes the Day of the Week

```python
day=df.groupby(by=['day of week','hour']).count()['Reason']
new_df=day.unstack().head()
new_df
```

```python
plt.subplots(figsize=(12,6))
sns.heatmap(new_df,cmap='viridis')
plt.tight_layout()
```

 Creating a clustermap using this DataFrame.

```python
plt.subplots(figsize=(12,6))
sns.clustermap(new_df,cmap='viridis')
```
