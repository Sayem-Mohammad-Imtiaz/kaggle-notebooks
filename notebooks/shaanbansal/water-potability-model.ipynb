```python
df2=df1.drop('Trihalomethanes',axis='columns')
```

```python
input=df2.drop('Potability',axis='columns')
target=df2['Potability']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,target)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn import tree
ref=tree.DecisionTreeClassifier()
```

```python
ref.fit(x_train,y_train)
```

```python
ref.score(x_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
data={
    'Algoithm':['LinearRegression','Randomforest','DecisionTree','LogisticRegression'],
    'score':[reg.score(x_test,y_test),model.score(x_test,y_test),ref.score(x_train,y_train),lr.score(x_train,y_train)]
}
loo1=pd.DataFrame(data)
loo1
```
