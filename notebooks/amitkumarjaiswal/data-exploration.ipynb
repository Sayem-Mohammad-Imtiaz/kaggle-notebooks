**Country Distribution**

Visualise how loans are distributed by country.

```python
%matplotlib inline
import plotly.graph_objs as go
import plotly.offline as py
py.init_notebook_mode(connected=True)
```

```python
large_counts = kiva_df['country'].value_counts().head(50)
trace = go.Bar(
    y=large_counts.index[::-1],
    x=large_counts.values[::-1],
    orientation='h'
)
layout = go.Layout(
    title='Country Distribution of Loans',
    width=700,
    height=1000
)

figure = go.Figure(data=[trace], layout=layout)
py.iplot(figure, filename="CountryLoan")
```

**Plotly ColorMap for Countries**

```python
map_df = kiva_df['country'].value_counts().to_frame().reset_index()
map_df.columns = ['country', 'loan_count']
```

```python
map_plot_data = [
    {
        'type': 'choropleth',
        'locations': map_df['country'],
        'locationmode': 'country names',
        'z': map_df['loan_count'],
        'text': map_df['country'],
        'autocolorscale': True
    }
]
map_plot_layout = {
    'title': 'Loans By Country'
}
map_plot_fig = {'data': map_plot_data, 'layout': map_plot_layout}
py.iplot(map_plot_fig, filename='loan_world_map')
```

**Explore Sector Distribution**

```python
def distribution_by_bar(series_df, title):
    trace = go.Bar(
        y=series_df.index[::-1],
        x=series_df.values[::-1],
        orientation='h'
    )
    layout = go.Layout(
        title=title,
        width=700,
        height=1000
    )

    figure = go.Figure(data=[trace], layout=layout)
    py.iplot(figure, filename=title)
```

```python
distribution_by_bar(kiva_df['sector'].value_counts(), 'LoanBySector')
```
