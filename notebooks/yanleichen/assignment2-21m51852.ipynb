# create a global map for COVID Vaccinations (Percentage of the population that are fully vaccinated)

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header = 0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
print(df.columns)
fig = px.choropleth(df, locations = 'country', locationmode = 'country names', color = 'people_fully_vaccinated_per_hundred',hover_name = 'country',animation_frame = 'date',color_continuous_scale = 'Plasma',range_color = (0.0,40.0))
fig.update_layout(title_text="Percentage of the country's total population that are fully vaccinated",title_x = 0.45)
fig.show()

plotly.offline.plot(fig,filename = 'Vaccniation.html')
```


1.The reason why I choose the range of color to be 0:40 that it still be an early vaccination stage in world scale which means that most of countries will lack of the vaccine so the number of fully vaccination even in USA won't be much high.

2.I noticed that some wealthy countries in western (USA and west Europe) had started to vaccinate people firstly. Recently, USA has achieved to nearly 40 people fully vaccination per hunderd people which is a very gerat result. However, those countries which are relatively low-income countries in Africa and middle-east still waiting for their first does of vaccine from COVAX which is the COVID-19 vaccine delivery project from WHO. There will be a serious problem in achieving herd immunity soon. Some developed countries should do better on helping the disadvantaged countries, especially the country who stock up on vaccines that are disproportionate to the country’s population.

# create a global map of earthquake occurrenes worldwide 

```python
df = pd.read_csv('../input/2021-recent-earthquake/query.csv',header = 0)
df.index = pd.to_datetime(df['time'])
df['time'] = df.index.strftime('%Y-%m-%d %H:00:00')
df = df.sort_index()
fig = px.scatter_geo(df,lat = 'latitude',lon ='longitude', color = 'mag',animation_frame ='time',color_continuous_scale ='rainbow',range_color = (1.0,8.0))
fig.show()
```


1. The range of magnitude I choosed is 1.0 to 8.0 casuse I want to check if there were some huge earthquakes happened recently. 
2. I noticed that there were 4 cases of Magnitude 5+ earthquakes happened. Besides the earthquakes happended in North Pacific Ocean, Panama and North Pacific Ocean, 92 km east of Hachinohe, Aomori, Japan, a 5.3 earthquake happened 110 km west of Tupiza, Sud Chichas, Departamento de Potosi, Bolivia, on 1 Jun 8:42 am (GMT -4) and a mag. 5.1 earthquake happened 88 km east of San José de Jachal, Departamento de Jachal, San Juan, Argentina, on 31 May 11:48 pm (GMT -3) .
