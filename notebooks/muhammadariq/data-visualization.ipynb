```python
df_housing = pd.read_csv('../input/california-housing-prices/housing.csv')
df_housing.head()
```

```python
plt.figure(figsize=(9,9))
sns.scatterplot(x='longitude', y='latitude', hue='median_house_value', size='population', data=df_housing, palette='mako')
plt.show()
```

Houses in the West or by the sea are more expensive than houses in other locations in California.

```python
df_housing.plot(
kind='scatter',
x='longitude',
y='latitude',
alpha=0.4,
s=df_housing['population']/100,
label='population',
c='median_house_value',
cmap=plt.get_cmap('jet'),
colorbar=True,
figsize=(12,9))
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(9,9))
sns.scatterplot(x='longitude', y='latitude', hue='ocean_proximity', data=df_housing, alpha=0.8)
plt.show()
```

```python
fig = plt.figure(figsize=(12,12))
ax = fig.add_subplot(projection='3d')

ax.scatter(df_housing['housing_median_age'], df_housing['median_income'], df_housing['median_house_value'], c=df_housing['median_house_value'], alpha=0.4)
ax.set_xlabel('median age')
ax.set_ylabel('median income')
ax.set_zlabel('median house value')
```

```python
plt.figure(figsize=(9,5))
sns.scatterplot(x='median_income', y='median_house_value', data=df_housing, alpha=0.2)
plt.show()
```
