```python
import pandas as pd
df = pd.read_csv('../input/mnist-in-csv/mnist_train.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
X = df.drop('label',axis=1).values
Y = df[['label']].values
```

```python
import matplotlib.pyplot as plt
plt.imshow(X[61].reshape(28,28))
plt.show()
```

```python
#Divide into train & test
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y)
```

```python
xtrain.shape[1]
ytrain.shape
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
```

```python
from keras import layers
from keras import models
```

```python
from keras.utils import np_utils
```

```python
ytrain_h = np_utils.to_categorical(ytrain)
ytest_h = np_utils.to_categorical(ytest)
```

```python
xtrain_n = xtrain/xtrain.max()
xtest_n = xtest/xtest.max()
```

```python
plt.imshow(xtrain_n[78].reshape(28,28))
plt.show()
```

```python
model = models.Sequential()
model.add(layers.Dense(200 , activation = 'relu' , input_shape=(xtrain.shape[1],)))
model.add(layers.Dense(200 , activation = 'relu'))
model.add(layers.Dense(10,activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy',
             optimizer = 'sgd',
             metrics = ['Accuracy'])
```

```python
model.fit(xtrain_n,ytrain_h,epochs=10)
```

**Functional API of Keras**

```python
Inp = layers.Input(shape=(784,))
h1 = layers.Dense(200,activation='relu')(Inp)
h2 = layers.Dense(100,activation='relu')(h1)
h3 = layers.Dense(100,activation='relu')(h1)
output = layers.Dense(10,activation='sigmoid')(h3)
```

```python
f_model = models.Model(Inp,output)
```

```python
f_model.summary()
```

```python
f_model.compile(loss = 'categorical_crossentropy',
               optimizer = 'sgd',
               metrics = ['Accuracy'])
```

```python
f_model.fit(xtrain_n,ytrain_h,epochs=20)
```
