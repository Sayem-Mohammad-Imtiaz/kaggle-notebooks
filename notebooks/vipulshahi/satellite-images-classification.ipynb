```python
import pandas as pd
X = pd.read_csv('../input/deepsat-sat4/X_test_sat4.csv').values
```

```python
Y = pd.read_csv('../input/deepsat-sat4/y_test_sat4.csv').values
```

```python
print(X.shape)
print(Y.shape)
```

```python
#Barren land, Trees, GrassLand, Others
def label(x):
    if Y[x][0] == 1:
        return 'Barren land'
    elif Y[x][1] == 1:
        return 'Trees'
    elif Y[x][2] == 1:
        return 'Grassland'
    elif Y[x][3] == 1:
        return 'Others'
```

```python
import numpy as np
Yt = []
for i in Y:
    Yt.append(np.argmax(i))
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Yt)
```

```python
from sklearn.linear_model import LogisticRegression
lmodel = LogisticRegression()
```

```python
ytest = np.array(ytest).reshape(25000,1)
ytest.shape
```

```python
lmodel.fit(xtrain,ytrain)
```

```python
print(xtrain.shape)
print(xtest[10].shape)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(ytest,lmodel.predict(xtest))
```

```python
print(lmodel.score(xtrain,ytrain))
print(lmodel.score(xtest,ytest))
```
