```python

# Importing the Libraries
import numpy as np # Used for any type of Mathematical operation
import matplotlib.pyplot as plt
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
# Importing the Datasets
haberman_data= pd.read_csv ('../input/habermans-survival-data-set/haberman.csv')  
print(haberman_data.head())
```

```python
import seaborn as sns
sns.heatmap(haberman_data.corr())
```

```python
# Extracting Independent Variable
X= haberman_data.iloc[:,0:2].values
# Extracting Dependent Variable
Y= haberman_data.iloc[:,3].values   # iloc is method of panda, 1st(:)represent all rows and 2nd(:) represent all columns , -1 leaving the last row
```

```python
# Spliting the data into train and test set:

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(X,Y, test_size=0.25 , random_state=0)
```

```python
#import feature scaling
from sklearn.preprocessing import StandardScaler
St_x=StandardScaler()
X_train = St_x.fit_transform(X_train)
X_test = St_x.transform(X_test)
```

```python
#Fitting logistic Regression model to training 

from sklearn.linear_model import LogisticRegression
model = LogisticRegression() # instance of a class
model.fit (X_train , Y_train)
```

```python
# Predicting the test set Results for Logistic:
pred_log = model.predict(X_test)
pred_log
```

```python
#import DecisionTreeClassifier  

from sklearn.tree import DecisionTreeClassifier  
model= DecisionTreeClassifier (criterion='entropy', random_state=0)
model.fit(X_train , Y_train)
```

```python

# Predicting the test set Results for DecisionTrees:
pred = model.predict(X_test)
pred
```

```python
# import KNN Classifier:
from sklearn.neighbors import KNeighborsClassifier
model= KNeighborsClassifier(n_neighbors=7, metric='minkowski', p=2 )
model.fit(X_train , Y_train)
```

```python

# Predicting the test set Results for DecisionTrees:
pre = model.predict(X_test)
pre
```

```python
# importing confusion metrics
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(Y_test, pred_log) 
cm
```

```python
# importing accuracy functions
from sklearn.metrics import accuracy_score 
Score=accuracy_score(Y_test, pred_log)
Score
```
