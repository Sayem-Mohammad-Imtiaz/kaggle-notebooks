```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

selected_country_1='Nepal'
selected_country_2='Nepal'

df1=pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1=df1[df1['country']==selected_country_1]
df2=pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2=df2[df2['country']==selected_country_2]
df1.index=pd.to_datetime(df1['date'])
df2.index=pd.to_datetime(df2['date'])
```

# the situation of COVID-19 in Nepal
the below figure shows that the relation between daily comfirmed cases and the situation of vaccinations.
the number of case of COVID-19 was stable from March 2020 to around the beginning of July 2020. However the number of it has shot up from April 2021. According to CNN news, this factor is new coronavirus variant from India. now the number of it is 42th most in the world.
and about vaccinations, Assisting by China enable to increase the number of vaccination. the rate of people vaccinated per a hundred is 2.11. on the other hand, that in Japan is 2.31. from this figure you would think the situation of Nepal is not bad. but region hospitals are straining to keep up with COVID-19 hospitalizations.

```python
fig,ax=plt.subplots(figsize=(10,5))
ax.plot(df1.index,df1['daily_new_cases'])
ax1=ax.twinx()
ax1.plot(df2.index,df2['daily_vaccinations'],'r')
plt.show()
```

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_new_cases = go.Scatter(x=df1.index,y=df1['daily_new_cases'],yaxis='y1',name='Daily confirmed cases')
daily_vaccinations = go.Scatter(x=df2.index,y=df2['daily_vaccinations'],yaxis='y2',name='Daily vaccinations')

layout_obj=go.Layout(title='COVID vs. Vaccines',xaxis=dict(title='Date'),yaxis=dict(title='Daily cases'),yaxis2=dict(title='Vaccines'))
fig=go.Figure(data=[daily_new_cases,daily_vaccinations],layout=layout_obj)
iplot(fig)
```
