```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
sns.set()
```

```python
df=pd.read_csv('/kaggle/input/heart-disease-uci/heart.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
# looking for any null values
df.isnull().sum()
```

```python
# count the no. of target
df.target.value_counts()
```

```python
# lets plot the graph
sns.countplot(x='target',data=df,palette="bwr")
plt.plot()
```

```python
# lets find out the percentage 
countNoDisease = len(df[df.target == 0])
countHaveDisease = len(df[df.target == 1])
print("Percentage of Patients Haven't Heart Disease: {:.2f}%".format((countNoDisease / (len(df.target))*100)))
print("Percentage of Patients Have Heart Disease: {:.2f}%".format((countHaveDisease / (len(df.target))*100)))
```

```python
# ploting the graph for no of male and female
sns.countplot(x='sex', data=df, palette="mako_r")
plt.xlabel("Sex (0 = female, 1= male)")
plt.show()
```

```python
#lets find out the percentage  
countfemale=len(df[df.sex==0])
male=len(df[df.sex==1])
print("Percentage of Patients are female: {:.2f}%".format((countfemale / (len(df.sex))*100)))
print("Percentage of Patients are male: {:.2f}%".format((male/ (len(df.sex))*100)))
```

```python
df.groupby('target').mean()
```

```python
pd.crosstab(df.age,df.target).plot(kind="bar",figsize=(20,6))
plt.title('Heart Disease Frequency for Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.savefig('heartDiseaseAndAges.png')
plt.show()
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
df['target'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Rate of Heart Disease')
ax[0].set_ylabel('Count')
sns.countplot('target',data=df,ax=ax[1],order=df['sex'].value_counts().index)
ax[1].set_title('Rate of Heart Disease')
plt.show()
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
df['sex'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Heart Disease by sex')
ax[0].set_ylabel('Count')
sns.countplot('sex',data=df,ax=ax[1],order=df['sex'].value_counts().index)
ax[1].set_title('Heart Disease by Sex')
plt.show()
```

```python
f,ax=plt.subplots(1,2,figsize=(18,8))
df['cp'].value_counts().plot.pie(explode=[0,0.05,0.05,0.05],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title('Chest Pain Type')
ax[0].set_ylabel('Count')
sns.countplot('cp',data=df,ax=ax[1],order=df['cp'].value_counts().index)
ax[1].set_title('Chest Pain Type')
plt.show()
```

```python
#Creating subplots
plt.figure(figsize=(15, 12))
#Subplot1 
plt.subplot(2,2,1)
plt.title('Distribution of Age')
sns.distplot(df['age'], rug = True)

#Subplot2
plt.subplot(2,2,2)
plt.title('Distribution of Resting Blood Pressure')
sns.distplot(df['trestbps'], rug = True)

#Subplot3
plt.subplot(2,2,3)
plt.title('cholesterol')
sns.distplot(df['chol'], rug = True)


#Subplot4
plt.subplot(2,2,4)
plt.title('Distribution of Max Hear Rate')
sns.distplot(df['thalach'], rug = True)
plt.ioff()
#plt.show()
```

```python
fig,ax = plt.subplots(nrows=4, ncols=2, figsize=(18,18))
plt.suptitle('Violin Plots',fontsize=24)
sns.violinplot(x="cp", data=df,ax=ax[0,0],palette='Set3')
sns.violinplot(x="trestbps", data=df,ax=ax[0,1],palette='Set3')
sns.violinplot (x ='chol', data=df, ax=ax[1,0], palette='Set3')
sns.violinplot(x='fbs', data=df, ax=ax[1,1],palette='Set3')
sns.violinplot(x='oldpeak', data=df, ax=ax[2,0], palette='Set3')
sns.violinplot(x='thalach', data=df, ax=ax[2,1],palette='Set3')
sns.violinplot(x='exang', data=df, ax=ax[3,0],palette='Set3')
sns.violinplot(x='age', data=df, ax=ax[3,1],palette='Set3')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot("sex","trestbps",hue="target", data=df,split=True)
plt.subplot(2,2,2)
sns.violinplot("sex","chol",hue="target", data=df,split=True)
plt.subplot(2,2,3)
sns.violinplot("sex","thalach",hue="target", data=df,split=True)
plt.subplot(2,2,4)
sns.violinplot("sex","fbs",hue="target", data=df,split=True)
plt.ioff()
plt.show()
```

```python
pd.crosstab(df.age,df.target).plot(kind='bar',figsize=(20,6))
plt.title('Heart Disease Vs Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.savefig('HeartDiseaseAge.png')
plt.ioff()
```

```python
plot = df[df.target == 1].age.value_counts().sort_index().plot(kind = "bar", figsize=(15,5), fontsize = 15)
plot.set_title("Heart disease: Age distribution", fontsize = 15)
plt.ioff()
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.scatter(x=df.age[df.target==1],y=df.thalach[df.target==1],c='red')
plt.scatter(x=df.age[df.target==0],y=df.thalach[df.target==0],c='green')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.legend(['Disease','No Disease'])

plt.subplot(1,3,2)
plt.scatter(x=df.age[df.target==1],y=df.chol[df.target==1],c='red')
plt.scatter(x=df.age[df.target==0],y=df.chol[df.target==0],c='green')
plt.xlabel('Age')
plt.ylabel('Cholesterol')
plt.legend(['Disease','No Disease'])

plt.subplot(1,3,3)
plt.scatter(x=df.age[df.target==1],y=df.trestbps[df.target==1],c='red')
plt.scatter(x=df.age[df.target==0],y=df.trestbps[df.target==0],c='green')
plt.xlabel('Age')
plt.ylabel('Resting Blood Pressure')
plt.legend(['Disease','No Disease'])

plt.tight_layout()
plt.ioff()
```
