```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
pd.set_option('display.max_rows', None)
from plotly.subplots import make_subplots
import seaborn as sns
import datetime


import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
data = pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/covid_19_data.csv')
data[["Confirmed","Deaths","Recovered"]] =data[["Confirmed","Deaths","Recovered"]].astype(int)
data['Active_cases'] = data['Confirmed'] - data['Deaths'] - data['Recovered']
data.head()
```

```python
pais = data [(data['Country/Region'] == 'Colombia') ].reset_index(drop=True)
ultima = pais[pais['ObservationDate'] == max(pais['ObservationDate'])].reset_index()
total= ultima.groupby(["Country/Region"])["Active_cases","Deaths","Recovered"].sum().reset_index().reset_index(drop=True)
total
```

```python
labels = ["Active cases","Recovered","Deaths"]
values = total.loc[0, ["Active_cases","Recovered","Deaths"]]
df = px.data.tips()
fig = px.pie(total, values=values, names=labels, color_discrete_sequence=['royalblue','darkblue','green'], hole=0.5)
fig.update_layout(
    title='Total de casos Colombia : '+str(total["Active_cases"][0]),
)
fig.show()

```

```python
data["Province/State"]= data["Province/State"].fillna('Unknown')
```

```python
Data= data[data['ObservationDate'] == max(data['ObservationDate'])].reset_index()
Data_world = Data.groupby(["ObservationDate"])[["Confirmed","Active_cases","Recovered","Deaths"]].sum().reset_index()
Data_world
```

```python
labels = ["Active cases","Recovered","Deaths"]
values = Data_world.loc[0, ["Active_cases","Recovered","Deaths"]]
fig = px.pie(Data_world, values=values, names=labels,color_discrete_sequence=['rgb(77,146,33)','rgb(69,144,185)','rgb(77,77,77)'],hole=0.7)
fig.update_layout(
    title='Total cases : '+str(Data_world["Confirmed"][0]),
)
fig.show()
```

```python
data_over_time= data.groupby(["ObservationDate"])[["Confirmed","Active_cases","Recovered","Deaths"]].sum().reset_index().sort_values("ObservationDate",ascending=True).reset_index(drop=True)
data_over_time
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=data_over_time.index, y=data_over_time['Confirmed'],
                    mode='lines',
                    name='Confirmed cases'))


fig.update_layout(
    title='Evolution de casos confirmados en el mundo',
        template='plotly_white',
      yaxis_title="Confirmed cases",
    xaxis_title="Days",

)

fig.show()
```
