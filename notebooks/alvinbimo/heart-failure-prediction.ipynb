# Import Library

```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn
 
 
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import sklearn.metrics as metrics
from matplotlib import pyplot as plt
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import classification_report
```

# Load Data

```python
df = pd.read_csv('../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv')
df
```

# Data Preprocessing

```python
df = df.drop(['time','sex'],axis='columns')
df = df.drop('age',axis = 'columns')
```

```python
df.columns
```

```python
df.info()
```

```python
x = df.drop(columns='DEATH_EVENT')
y = df['DEATH_EVENT']
```

```python
x.columns = ['anaemia', 'creatinine_phosphokinase', 'diabetes', 
             'ejection_fraction', 'high_blood_pressure', 'platelets',
             'serum_creatinine', 'serum_sodium', 'smoking']
y.columns = ['1', '0']
```

# Data Splitting

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100)
```

# Making Model

```python
clf = DecisionTreeClassifier(criterion='entropy', max_depth=5, random_state=160, min_samples_leaf=20, min_samples_split=9)
clf
```

```python
clf.fit(x_train, y_train)
```

# Predicting

```python
y_pred = clf.predict(x_test)
print(y_pred)
```

```python
cm = np.array(confusion_matrix(y_test, y_pred))
cm
```

```python
print('Accuration: ', metrics.accuracy_score(y_test, y_pred))
```

# Model Visualization

```python
fig = plt.figure()
sna = sns.heatmap(cm, annot=True, xticklabels=y.columns, yticklabels=y.columns)
fig.show()
```

```python
figl = plt.figure(figsize=(30,30))
tree.plot_tree(clf, feature_names = x.columns, class_names = y.columns)
figl.show()
```
