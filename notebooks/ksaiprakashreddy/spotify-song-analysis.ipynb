Import Packages and load dataset

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import plotly.express as px
df=pd.read_csv("../input/top50spotify2019/top50.csv",encoding="ISO-8859-1")
df.head()
```

Plot the Treemap for Genre and Artist Name according to Popularity

```python
fig = px.treemap(df,path=['Genre', 'Artist.Name'],values=df['Popularity'])
fig.show()
```

Droping the 'Unnamed: 0' column, as it doesn't contribute for the Analysis

```python
df.drop(['Unnamed: 0'],axis = 1, inplace = True)
```

Grouping the sum of Genre and ArtistName to visualize various artist belong to specific genre and values related to other variables

```python
relation = df.groupby(['Genre', 'Artist.Name']).sum()
relation
```

Plotting the Histogram for the Popularity of songs

```python
plt.figure(figsize = (10,7))
sns.set_style("whitegrid")
sns.distplot(df['Popularity'])
plt.show()
```

Historgram for Dancability of songs

```python
plt.figure(figsize=(10,7))
sns.distplot(df['Danceability'], kde=False, bins=15,color='red')
plt.show()
```

plotting the heatmap for correlation of various attributes using pearson method

```python
# heatmap of the correlation 

correlation=df.corr(method='pearson')
plt.figure(figsize=(10,10))
plt.title('Correlation heatmap',fontsize=20)
sns.heatmap(correlation,annot=True, center = 1)
plt.show()
```
