```python
data1 = data[['satisfaction_level', 'average_montly_hours', 'promotion_last_5years', 'salary']]
```

```python
dummies = pd.get_dummies(data.salary)
dummies
```

Since, Salary component is not a numerical data so getting dummy data for salary.

```python
main_data=pd.concat([data1,dummies],axis='columns')
main_data
```

Concatenating the two dataframes and droping salary column becasue we don't need it for our model.

```python
main_data.drop(['salary'],axis='columns',inplace=True)
```

Now that our data have been converted into numerical encoded so we can apply data to our models. We'll use main_data for X and data.left for our y

```python
X= main_data
```

```python
y=data.left
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.20)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg = LogisticRegression()
```

```python
lg.fit(X_train,y_train)
```

```python
lg.predict(X_test)
```

```python
lg.score(X_test,y_test)
```
