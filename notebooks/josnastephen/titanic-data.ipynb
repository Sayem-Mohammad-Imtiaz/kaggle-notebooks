```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("../input/titanicdataset-traincsv/train.csv")
```

```python
df.head()
```

```python
df.drop(['Name','Ticket','Fare'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df['Embarked'].value_counts()
```

```python
df['Embarked'] = df['Embarked'].fillna('S')
```

```python
df.drop(['Cabin'],axis=1,inplace=True)
```

```python
df.head()
```

```python
sex=pd.get_dummies(df['Sex'],drop_first=True)
embark =pd.get_dummies(df['Embarked'],drop_first=True)
```

```python
df.drop(['Sex','Embarked'],axis=1,inplace=True)
```

```python
df =pd.concat([df,sex,embark],axis=1)
```

```python
df.head()
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Survived',data=df,palette='RdBu_r')
```

```python
X=df.drop(['Survived'],axis=1)
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

```

```python
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
pred=lr.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
print(classification_report(y_test,pred))
```

```python
dt =DecisionTreeClassifier()
```

```python
dt.fit(X_train,y_train)
```

```python
predd =dt.predict(X_test)
```

```python
confusion_matrix(y_test,predd)
```

```python
print(classification_report(y_test,predd))
```

```python
rf= RandomForestClassifier(n_estimators=500)
rf.fit(X_train,y_train)
```

```python
random_pred = rf.predict(X_test)
```

```python
confusion_matrix(y_test,random_pred)
```

```python
print(classification_report(y_test,random_pred))
```
