```python
import seaborn as sns
```

```python
#here we plot scatter plot with respect to class label
sns.set_style("whitegrid")
sns.FacetGrid(data, hue=3,height=6).map(plt.scatter,0,1).add_legend();
plt.show()
```

After observing the scatter plot we got to know that we cannot simply use column 0 and 1 to build a classifier model on this data set because data set is randomly distributed so we try to look at others combination

```python
#here we build pairplot and try to look is there any combinaton from features that can help us or not
plt.close()
sns.set_style("whitegrid")
sns.pairplot(data,hue=3,height=3)
```

By observing the above plot we found out that any two combination of features could not help us to easily classify the data set so we should try building more complex features

```python
#1-d plot or pdf
plt.close()
sns.set_style("whitegrid")
sns.FacetGrid(data, hue=3,height=6).map(sns.distplot,2).add_legend();
```

```python
#plotting pdf and cdf corresponding to column 0
counts,bin_edges=np.histogram(data[0],bins=10,density =True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```
