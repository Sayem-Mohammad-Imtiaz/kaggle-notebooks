```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date
df1 = pd.read_csv("/kaggle/input/solar-power-generation-data/Plant_1_Generation_Data.csv")
df1['Year'] = pd.to_datetime(df1['DATE_TIME']).dt.year
df1['month'] = pd.to_datetime(df1['DATE_TIME']).dt.month
df1['day'] = pd.to_datetime(df1['DATE_TIME']).dt.day
df1['weekday'] = pd.to_datetime(df1['DATE_TIME']).dt.weekday
print(df1)
```

```python
def print_time_frame(data, label):
    print(f"{label}  -  start:{data['DATE_TIME'].min()}   end:{data['DATE_TIME'].max()}")
print_time_frame(df1, 'Plant1 ')
```

```python
df2 = pd.read_csv("/kaggle/input/solar-power-generation-data/Plant_1_Weather_Sensor_Data.csv")
print(df2)
```

```python
dailymean= df1['DAILY_YIELD'].mean()
print("Mean of Daily yield is",dailymean)
max_am=df2['AMBIENT_TEMPERATURE'].max()
print("Maximum ambient temperature is",max_am)
max_md=df2['MODULE_TEMPERATURE'].max()
print("Maximum module temperature is",max_md)
min_am=df2['AMBIENT_TEMPERATURE'].min()
print("Minimum ambient temperature is",min_am)
min_md=df2['MODULE_TEMPERATURE'].min()
print("Minimum module temperature is",min_md)
```

```python
d=df1.groupby('day')['TOTAL_YIELD']
f2=d.sum().plot.bar( width=0.4, fontsize = 7, color=['red', 'blue', 'purple', 'green', 'orange','yellow'], edgecolor='black', align='center',linewidth=2)
plt.xlabel('Weekday',fontsize=10)
plt.ylabel('Total Yeild',fontsize=10)
plt.title('Daily Total Yield',fontsize=10)
plt.show()
```

```python
d=df1.groupby('weekday')['TOTAL_YIELD']
f2=d.sum().plot.bar( width=0.4, fontsize = 7, color=['red', 'blue', 'purple', 'green', 'orange'], edgecolor='black', align='center',linewidth=2)
plt.xlabel('Day',fontsize=10)
plt.ylabel('Total Yeild',fontsize=10)
plt.title('Daily Total Yield',fontsize=10)
f2.set_xticklabels(('Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'))
plt.show()
```
