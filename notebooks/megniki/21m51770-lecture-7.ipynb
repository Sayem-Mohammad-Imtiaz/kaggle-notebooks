# Global Map of COVID Vaccinations (Percent of Fully Vaccinated People)

Below, I have illustrated the number of fully vaccinated people per 100 in every country on a map. I have selected this information because I believe it best represents the progress of vaccinations and therefore the readiness of each country to exit out of the pandemic (people must be fully vaccinated to be as protected against COVID as possible).

I have updated the figure so that the animation moves at the speed of my preference.

```python
import pandas as pd
import numpy as np
import plotly.express as px

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
fig = px.choropleth(df,locations='country',locationmode='country names',color='people_fully_vaccinated_per_hundred',hover_name='country',animation_frame='date',color_continuous_scale='Pinkyl',range_color=(0.0,50.0))
fig.update_layout(title_text='Percent of Fully Vaccinated People')
fig.update_layout(title_font_size=20,title_x=0.40,title_font_color='coral')
fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 30
fig.layout.updatemenus[0].buttons[0].args[1]['transition']['duration'] = 5
fig.show()
```

# Global Map of Earthquakes Between 12:00 UTC on January 12, 2010  and 11:59 UTC on January 13, 2010 (2010 Haiti Earthquake)

Below, I have created an hourly animation of global earthquake locations between 12:00 UTC on January 12, 2010 and 11:59 UTC on January 13, 2010 (The animation spans over 24 hours).

A magnitude 7.0 M earthquake occurred inland in Haiti at 16:43 local time or 21:43 UTC on January 12, 2010. This earthquake can be observed in the animation below (In the animation, it appears as occuring at 21:00 because it has been converted to hourly data).

According to DesRoches et al. (2011), the earthquake left more than 316,000 people dead or missing and 300,000 injured. Over 300,000 homes collapsed or were critically damaged, and 1.3 million people became homeless. At the time, Haiti ranked 145th out of 169 countries on the UN Human Developement Index, and had little experience of large earthquakes in modern times. As a result, the country was unprepared for the earthquake and the impact was devastating. According to UNICEF (2020), 10 year after the earthquake, the country is still on the road to recovery and development and faces multiple socioeconomic crises.

```python
df2 = pd.read_csv('../input/january2010-earthquake/earthquake_jan2010.csv',header=0)
df2.index = pd.to_datetime(df2['time'])

df2['time'] = df2.index.strftime('%Y-%m-%d %H:00:00')
fig2 = px.scatter_geo(df2,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='matter',range_color=(1.0,7.0))
fig2.update_layout(title_text='Earthquakes Jan. 12 12:00 UTC - Jan. 13 11:59 UTC, 2010')
fig2.update_layout(title_font_size=14,title_x=0.50,title_font_color='darkmagenta')
fig2.show()
```
