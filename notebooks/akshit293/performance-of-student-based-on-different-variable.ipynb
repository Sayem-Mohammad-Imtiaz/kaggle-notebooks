# Performance Of Student based on gender

```python
#almost equal number of male and female candidates
data['gender'].value_counts()
```

```python
sns.catplot(x='gender',y='math score',data=data,kind='swarm')
sns.catplot(x='gender',y='math score',data=data,kind='point')
```

* There are many female candidate outliers in math score(expecially scoring marks less than 40).
* Average marks scored by female candidates is less than that of male candidates.
* Number of female candidate scoring less than 40 is more than that of male candidate scoring less than 40.

```python
sns.catplot(x='gender',y='reading score',data=data,kind='swarm')
sns.catplot(x='gender',y='reading score',data=data,kind='point')
```

* Both male and female candidate have outliers.
* Average reading marks scored by female candidates is more than that of male candidates.
* Number of female candidate scoring 100 is more than that of male scoring 100.

```python
sns.catplot(x='gender',y='writing score',data=data,kind='swarm')
sns.catplot(x='gender',y='writing score',data=data,kind='point')
```

* Both male and female candidate have outliers.
* Average reading marks scored by female candidates is more than that of male candidates.
* Number of female candidate scoring 100 is more than that of male scoring 100.

# Conclusion

1. Group of male will collectively score more marks in math as compared to the of goup of female.
2. Group of male will collectively score less marks in reding as compared to the of goup of female.
3. Group of male will collectively score less marks in writing as compared to the of goup of female.

# Performance of student based on race/ethnicity 

```python
#imbalanced class
data['race/ethnicity'].value_counts()
```

```python
sns.catplot(y='math score',x='race/ethnicity',data=data,kind='swarm')
sns.catplot(y='math score',x='race/ethnicity',data=data,kind='point')
```

* group E candidate has highest average math score(although number of group E candidates are 140)
* group A candidate has lowest average math score(group A is the smallest group) 

```python
sns.catplot(y='reading score',x='race/ethnicity',data=data,kind='swarm')
sns.catplot(y='reading score',x='race/ethnicity',data=data,kind='point')
```

just like math score
* group E candidate has highest average reading score(although number of group E candidates are 140)
* group A candidate has lowest average reading score(group A is the smallest group) 

```python
sns.catplot(y='writing score',x='race/ethnicity',data=data,kind='swarm')
sns.catplot(y='writing score',x='race/ethnicity',data=data,kind='point')
```

**Checking median score of all groups**

```python
groupA=data[data['race/ethnicity']=='group A']
groupB=data[data['race/ethnicity']=='group B']
groupC=data[data['race/ethnicity']=='group C']
groupD=data[data['race/ethnicity']=='group D']
groupE=data[data['race/ethnicity']=='group E']

print("A\n",groupA.median(),"\n***********\n")
print("B\n",groupB.median(),"\n***********\n")
print("C\n",groupC.median(),"\n***********\n")
print("D\n",groupD.median(),"\n***********\n")
print("E\n",groupE.median(),"\n***********\n")
```

just like math score and reading score
* group E candidate has highest average writing score(although number of group E candidates are 140)
* group A candidate has lowest average writing score(group A is the smallest group) 

# Conclusion

* group E candidate will perform best among all other groups(collectively).
* group A candidate will perform worst among all other groups(collectively).
