# Taking A Look At Covid-19 Cases Situation In India

```python
dx.head(9).style.set_properties(**{'background-color':'black',
                                     'color': 'gold'})
```

```python
df=df[0:36]
```

```python
dx=dx.reset_index(drop=True)
```

```python
dxnew = pd.concat([df, dx], axis=1, join="inner")
```

```python
dxnew=dxnew.drop(columns=["Populations","State/UTs"],axis=1)
dxnew["States/UTs"]=dx["State/UTs"]
```

```python
dxnew.head()
```

```python
df1=dxnew.sort_values(by="Deaths",ascending=True)
df2=dxnew.sort_values(by="Deaths",ascending=False)
df3=dxnew.sort_values(by="Total Cases",ascending=True)
df4=dxnew.sort_values(by="Total Cases",ascending=False)
df5=dxnew.sort_values(by="Death Ratio (%)",ascending=True)
df6=dxnew.sort_values(by="Death Ratio (%)",ascending=False)
```
