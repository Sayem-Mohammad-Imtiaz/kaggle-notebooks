```python
import pandas as pd

df = pd.read_csv("../input/ipl-complete-dataset-20082020/IPL Ball-by-Ball 2008-2020.csv")
```

```python
df.drop(["batsman", "non_striker", "bowler","dismissal_kind","player_dismissed", "fielder", 
         "extras_type","batting_team","bowling_team", "non_boundary", "is_wicket", "extra_runs", 
         "total_runs"], axis = 1, inplace=True) # drop unnecessary columns 
```

```python
#match_id = int(input("Enter match id : ")) # example: 1237181, 335982, etc
match_id = 335984
#inning = int(input("Enter inning : "))
inning = 2
```

```python
df.drop(index= df[(df["batsman_runs"]!=4) & (df["batsman_runs"]!=6)].index, inplace=True) # keep 4 & 6s drop else
df['(over,ball)'] = df.apply(lambda x: "("+str(x['over'])+","+str(x['ball'])+")", axis=1) # create new column
df.sort_values(by=['id', 'inning', 'over', 'ball'], inplace=True) # Sort values because over and balls are random 
```

```python
df.head()
```

```python
import plotly.graph_objects as go

# https://stackoverflow.com/questions/56289777/coloring-scatter-plot-points-differently-based-on-certain-conditions
color = [ 
    0 if v == 4 else 1
    for v in df[(df['id']==match_id) & (df['inning']==inning)]['batsman_runs'].values
]

colorscale = [[0, 'red'], [1.0, 'green']]

fig = go.Figure(go.Scatter(
    x=df[(df['id']==match_id) & (df['inning']==inning)]['(over,ball)'], 
    y=df[(df['id']==match_id) & (df['inning']==inning)]['batsman_runs'],
    mode='markers',
    hovertemplate = "Score:%{y}: <br>(Over, Ball): %{x}",
    marker = {'color': color,
              'colorscale': colorscale,
              'size': 10
             },
))

fig.update_layout(
    title="Match Id: {} , Inning : {}".format(match_id, inning),
    yaxis = dict(
        tickmode = 'linear',
        tick0 = 4,
        dtick = 2,
        title = "Score",
    ),
    xaxis = dict(
        title = "(Over,ball)",
    )
)

fig.show()
```
