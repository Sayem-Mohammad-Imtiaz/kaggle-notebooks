```python

import plotly.express as px
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
df = pd.read_csv(r'/kaggle/input/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv')
```

```python
sns.heatmap(df.corr(),annot=True)
```

+ From the Correlation heat map, we can see the following :
    + Discharged cases and Total cases are linearly correlated
    + Deaths, Active are also near linearly related to Total cases
    + Deaths, Discharged are also linearly related
    

```python
plt.figure(figsize=(10,10))
sns.barplot(data=work_df,x='Active',y='State/UTs')
```

+ Kerala is having highest number of active cases followed by Maharashtra and Karnataka

```python
plt.figure(figsize=(10,10))
sns.barplot(data=work_df,x='Total Cases',y='State/UTs')
```

+ Maharshtra accounts for the highest number of total cases(since covid-19 beginning) followed by Kerala,Karnataka,Tamilnadu.

```python
px.bar(df,y="State/UTs",x="Deaths",color="Death Ratio (%)",title="Total Cases each State Wise")
```

+ Even though Kerala accounts most active cases, death ratio is low and punjab accounts for highest death ratio.

```python
px.sunburst(work_df,color='Death Ratio (%)',values='Total Cases',path=['State/UTs'],height=500)
```

+ Another easy way of visualizing the  Death Ratio and Total cases through interactive pie chart

```python
px.sunburst(work_df,color='Total Cases',values='Active',path=['State/UTs'],height=500)
```

Acive Cases(size) and Total Cases(color) representation shows that kerala leads in active cases and Maharashtra in the later.

```python
px.sunburst(work_df,color='Total Cases',values='Active Ratio (%)',path=['State/UTs'],height=500,title='Current Active case scenario of each state')
```

+ Here Is another information about Active Ratio(size) and total cases
    + Even though Mizoram does not have higher total cases, it has higher active ratio indicating lower discharges which is an important information to consider.

```python
px.sunburst(work_df,color='Total Cases',values='Discharge Ratio (%)',path=['State/UTs'],height=500,title='Current Active case scenario of each state')
```

+ The above pie chart shows that Mizoram has lowest recovery(discharge ratio) of all the states

```python
fig = px.pie(df,values='Active',names='State/UTs',title='Share of Active Cases')
fig.show()
```
