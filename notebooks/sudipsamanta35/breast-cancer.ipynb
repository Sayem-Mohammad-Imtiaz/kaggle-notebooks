```python
#splitting in two
X_data=data.iloc[:,2:31].values
Y_data=data.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_data,Y_data,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

```python
#Difference model

def models(X_train,Y_train):
    
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression()
    log.fit(X_train,Y_train)
    
    from sklearn.tree import DecisionTreeClassifier
    dtree=DecisionTreeClassifier()
    dtree.fit(X_train,Y_train)
    
    from sklearn.ensemble import RandomForestClassifier
    ran_for=RandomForestClassifier()
    ran_for.fit(X_train,Y_train)
    
    
    #print accuracy
 
    print('[1]Logistic Regression Training accuracy :',log.score(X_train,Y_train))
    print('[2]Decision Tree Training accuracy :',dtree.score(X_train,Y_train))
    print('[3]Random Forest Training accuracy :',ran_for.score(X_train,Y_train))
    return  log, dtree, ran_for
```

```python
model=models(X_train,Y_train)
```

```python
#test model accuracy on test data confusion matrix  without linear regression 


from sklearn.metrics import confusion_matrix

for i in range(len(model)):
    print("Model  : ", i+1)
    cm  =confusion_matrix(Y_test, model[i].predict(X_test))
    
    TP=cm[0][0]
    FP=cm[0][1]
    FN=cm[1][0]
    TN=cm[1][1]
    
    print(cm)
    print("Testing Accuracy =",(TP + TN) /( TP + TN + FN + FP))
    print()
```
