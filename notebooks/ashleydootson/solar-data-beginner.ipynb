```python
import matplotlib.pyplot as plt
import datetime as dt
```

```python
plant1_wsd = pd.read_csv('/kaggle/input/solar-power-generation-data/Plant_1_Weather_Sensor_Data.csv')
plant1_gen = pd.read_csv('/kaggle/input/solar-power-generation-data/Plant_1_Generation_Data.csv')
plant2_wsd = pd.read_csv('/kaggle/input/solar-power-generation-data/Plant_2_Weather_Sensor_Data.csv')
plant2_gen = pd.read_csv('/kaggle/input/solar-power-generation-data/Plant_2_Generation_Data.csv')
```

```python
plant1_wsd['DATE_TIME'] = pd.to_datetime(plant1_wsd['DATE_TIME'])
plant2_wsd['DATE_TIME'] = pd.to_datetime(plant2_wsd['DATE_TIME'])
plant1_gen['DATE_TIME'] = pd.to_datetime(plant1_gen['DATE_TIME'])
plant2_gen['DATE_TIME'] = pd.to_datetime(plant1_gen['DATE_TIME'])
```

```python
## Merged the corresponding data to provide overviw of both plant 1 & 2
all_plants_wsd = plant1_wsd.merge(plant2_wsd, how='outer')
all_plants_wsd['DD'] = all_plants_wsd['DATE_TIME'].dt.day
all_plants_wsd['MM'] = all_plants_wsd['DATE_TIME'].dt.month
print(all_plants_wsd.head())
all_plants_gen = plant1_gen.merge(plant2_gen, how='outer')
all_plants_gen['DD'] = all_plants_gen['DATE_TIME'].dt.day
all_plants_gen['MM'] = all_plants_gen['DATE_TIME'].dt.month
print(all_plants_gen.head())
```

```python
# Checking for any missing values
missing_values_wsd = all_plants_wsd.isna()
print(missing_values_wsd)
missing_values_gen = all_plants_gen.isna()
print(missing_values_gen)
```

```python
## show the mean daily yield by plant_id
p_yield = all_plants_gen['DAILY_YIELD'].groupby(all_plants_gen['PLANT_ID']).mean()
print(p_yield)
```

```python
## total irridation by day
total_irr = all_plants_wsd.pivot_table('IRRADIATION', ['DD', 'MM'], aggfunc='sum')
total_irr_sorted = total_irr.sort_values(['MM', 'DD'])
total_irr_sorted.plot()
```

```python
## max ambient and module temperature
max_ambient = all_plants_wsd['AMBIENT_TEMPERATURE'].groupby(all_plants_gen['PLANT_ID']).max()
print(max_ambient)
max_module = all_plants_wsd['MODULE_TEMPERATURE'].groupby(all_plants_gen['PLANT_ID']).max()
print(max_module)
```

```python
c_inv = all_plants_wsd['SOURCE_KEY'].groupby(all_plants_wsd['PLANT_ID']).count()
c_inv.plot(kind='bar')
plt.show()
```

```python
# max AC
max_ac = all_plants_gen.groupby([all_plants_gen['DATE_TIME']])['AC_POWER'].max()
print(max_ac)
min_ac = all_plants_gen.groupby([all_plants_gen['DATE_TIME']])['AC_POWER'].min()
print(min_ac)
```
