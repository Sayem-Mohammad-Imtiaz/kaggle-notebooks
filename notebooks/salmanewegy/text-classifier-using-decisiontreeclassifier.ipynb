```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
ArticlesDB = pd.read_csv('./../input/sms-spam-collection-dataset/spam.csv',usecols=['v1','v2'], encoding='latin-1')
```

```python
ArticlesDB.keys()
```

```python
ArticlesDB.NewsType.unique()
```

```python
x_train, x_test, y_train, y_test = train_test_split(ArticlesDB.drop('NewsType',axis=1), ArticlesDB['NewsType'], test_size=0.3, random_state=22)
```

```python
print("training examples=",len(x_train))
print("testing examples=",len(x_test))
print('total=',len(x_train)+len(x_test))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
x_trainV = vectorizer.fit_transform(x_train['Article'])
x_testV = vectorizer.transform(x_test['Article'])
x_train_tokens = vectorizer.get_feature_names()
print(x_train_tokens[-50:])
#print(x_trainV.toarray())
```

```python
DT = DecisionTreeClassifier(criterion="entropy", max_depth=14)
DT.fit(x_trainV, y_train)
y_pred = DT.predict(x_testV)
```

```python
from sklearn import metrics 
print("Accuracy: ", metrics.accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
plt.figure(figsize=(12, 6))
mat=confusion_matrix(y_test, y_pred)
sns.heatmap(mat.T,annot=True,xticklabels=y_train.unique(), yticklabels =y_train.unique())
plt.ylabel('true label')
plt.xlabel('predicted labels')
```
