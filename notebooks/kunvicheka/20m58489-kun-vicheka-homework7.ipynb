# Create a global map for COVID Vaccinations (Percentage of the poulation that are fully vaccinated)

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv', header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
print(df.columns)
fig = px.choropleth(df,locations='country',locationmode='country names',color='people_fully_vaccinated_per_hundred',hover_name='country',animation_frame='date',color_continuous_scale='turbo',range_color=(0.0,40.0))
fig.update_layout(title_text='Percentage of total population of the country that are fully vaccinated',title_x=0.45)
fig.show()

plotly.offline.plot(fig,filename='vaccinatons.html')
```

### Descript the visualization and the reason why we visualize
The column that we choose to visualize is the the total number of people fully vaccinated per hundred which is the ratio (in percent) between population fully immunized and total population up to the date in the country. Eventhough the number of vaccinations is increasing over the world, most vaccines require two or more doses to acquire immunity, and ony some vaccines, such as Janssen Pharmaceutical's, aim to confer immunity in a single dose. Therefore, it is really important to measure the number of people who have completed the required number of vaccinations when assessing overall progress. 

If we look at Israel, in terms of the number of people who have been vaccinated, Israel has inoculated 59.3 people per 100 people on 31 May 2021, which means that more than 1 in 2 people have already been vaccinated in Israel. At the same time, 29 May 2021, about 40.19 per 100 people in United States, and 37 per 100 in UK and Hungry, are fully vaccinated. 

# Create a global map of earthquake occurrences worldwide

```python
df = pd.read_csv('../input/202105recent-earthquake/query.csv',header=0)
df.index = pd.to_datetime(df['time'])
df['time'] = df.index.strftime('%Y-%m-%d %H:00:00')
df = df.sort_index()
fig = px.scatter_geo(df,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='turbo',range_color=(1.0,3.0) )
fig.show()
```

## Earthquake Occurrence Observation in Indonesia from 01 January 2021 to 03 May 2021

```python
df = pd.read_csv('../input/earthquake-from-01-january-to-current-time/query (1).csv',header=0)
df.index = pd.to_datetime(df['time'])
df['time'] = df.index.strftime('%Y-%m-%d %H:00:00')
df = df.sort_index()
fig = px.scatter_geo(df,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='turbo',range_color=(1.0,7.0) )
fig.show()
```

### Report on Earthquake in Indonesia from Januray to June 2021
> In January
On 14 January,reported 5.7 magnitude earthquake with 18m depth caused two houses and an emergency center damaged in Mamuju and one person sustained.
Followed up by 6.2 magnitude with 18m depth on 15 January in the same region, recorded as the **deadliest earthquake of 2021**, killed 105 people with 3 missings,and more than 3 300 others were injured.

> In February
A magnitude 6.2 offshore earthquake occurred in southwest of Bengkulu province, no initial report of damage and casualities. In north Maluku, another 5 magnitude happened which caused several damage facilities including 3 people injured.

> March 2021
No report of severe earthquake happened

> April 2021
On 10 April, a 6 magnitude with 67m depth earthquake occurred in East Java which caused 1000 building suffered significant damage and 9 people died with 104 injured. On 20 April, a 6.1 magnitude happened on North Sumatra with no report of casuality or injuring.

> May 2021
A 5.5 and 5.7 magnitude occured on 2 and 5 May with no significant damage. Another 6.7 magnitude in North Sumatra with no casuality or injuring report.

> June 2021
On 3 June 2021, an 6.1 magnitude occured with no report about casuality or injuring.
