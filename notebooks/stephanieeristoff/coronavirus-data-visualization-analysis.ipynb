# Time/Age Data Analysis
I decided to first analyze the Time/Age data, first printing out the first couple of entries to know what the labels of the Time/Age data to see how I could perform some data analysis.

```python
TimeAge.head()
```

I then decided to graph basic barplots for the age and the number of confirmed cases and the age and the number of deceased individuals. 

```python
sns.barplot(x='age', y = 'confirmed', data=TimeAge)
plt.title('Age Group and Confirmed Cases of Coronavirus')
sns.set_style("whitegrid")
plt.xlabel("Age Group")
plt.ylabel("Confirmed Coronavirus Cases")
plt.show()
sns.barplot(x='age', y = 'deceased', data=TimeAge)
plt.title('Age Group and Deceased Cases of Coronavirus')
plt.xlabel("Age Group")
plt.ylabel("Confirmed Coronavirus Deaths")
plt.show()
```

However, I realized that it would be more helpful to understand the ratio of deaths to confirmed cases to understand among what age groups is coronavirus most deadly.

```python
ratio_of_deaths_to_confirmed = TimeAge['deceased']/TimeAge['confirmed']
sns.barplot(TimeAge['age'], ratio_of_deaths_to_confirmed)
plt.title("Ratio of Deaths to Confirmed Coronavirus Cases and Age Group")
plt.xlabel("Age Group")
plt.ylabel("Ratio of Deaths to Confirmed Coronavirus Cases")
sns.set_style("whitegrid")
plt.show()
```

From this bar plot, we can see that the Ratio of Deaths to Confirmed cases and age group increases nearly exponentially as you increase the age by 10 year increments. This suggests that the older population is much more susceptible to death from coronavirus compared to younger populations.

This is expected as the older populations will generally be more susceptible to lung conditions, leading to death.

```python
sns.lineplot(x=TimeGender['date'], y=m_ratio_deceased_to_confirmed, label='Male')
sns.lineplot(x=TimeGender['date'], y = f_ratio_deceased_to_confirmed,label='Female')
plt.title("Deceased To Confirmed Sex Ratios Over Time")
plt.xlabel('Date')
plt.ylabel('Deceased To Confirmed Sex Ratios')
plt.show()
```

According to the above graph, men have a higher deceased to confirmed ratio. This indicates that men may more likely die when afflicted with the coronavirus.

# Time Province Data Visualization

Data Visualizations for the Time Province Data Set

```python
print(TimeProvince.head())
print(TimeProvince.tail())
```

Let's do some timeplots for each province. Let's first make an array of all the different provinces.

```python
provinces = TimeProvince.province.unique()
print(provinces)
```

I will now make a function that reads in a title and a dataset and will then create a timeplot for each province.

```python
def province_timeplot(pro):
    for i in pro:
        curr_y = TimeProvince['province'] == i
        sns.lineplot(x=curr_y['date'], y=curr_y['confirmed'])
        plt.title(i + " Province Coronavirus Confirmed Cases Time Plot")
        plt.xlabel('Date')
        plt.ylabel('Number of Confirmed Cases')
        plt.show()
```

Use the functions and the aforecreated array to create timeplots of each of the province's coronavirus confirmed cases:

```python
province_timeplot(provinces)
```
