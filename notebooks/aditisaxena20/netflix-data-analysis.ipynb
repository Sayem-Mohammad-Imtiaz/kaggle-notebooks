```python
df['country']=df['country'].fillna(df['country'].mode()[0])
df['date_added']=df['date_added'].fillna(df['date_added'].mode()[0])
df['rating']=df['rating'].fillna(df['rating'].mode()[0])
```

```python
msno.bar(df,figsize=(15,5),color='green',fontsize=10)
```

```python
df.drop(['cast','director'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.duplicated().sum()
```

```python
df['year']=df['date_added'].apply(lambda x: x.split(" ")[-1])
df['year'].head()
```

```python
df['month']=df['date_added'].apply(lambda x: x.split(" ")[0])
df['month'].head()
```

```python
ratings_ages = {
    'TV-PG': 'Older Kids',
    'TV-MA': 'Adults',
    'TV-Y7-FV': 'Older Kids',
    'TV-Y7': 'Older Kids',
    'TV-14': 'Teens',
    'R': 'Adults',
    'TV-Y': 'Kids',
    'NR': 'Adults',
    'PG-13': 'Teens',
    'TV-G': 'Kids',
    'PG': 'Older Kids',
    'G': 'Kids',
    'UR': 'Adults',
    'NC-17': 'Adults'
}
```

```python
df['target_age']= df['rating'].replace(ratings_ages)
```

```python
df['target_age'].head()
```

```python
df['country_main']=df['country'].apply(lambda x: x.split(",")[0])
df['country_main']
```
