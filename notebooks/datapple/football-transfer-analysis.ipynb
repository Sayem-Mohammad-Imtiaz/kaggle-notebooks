```python
df = pd.read_csv('/kaggle/input/top-250-football-transfers-from-2000-to-2018/top250-00-19.csv')
df.head()
```

```python
df.describe()
```

```python
df['Transfer_value'] = df['Transfer_fee'].apply(lambda x: "€{:.2f}M".format((x/1000000)))
```

#### ⚽Let's check in which age players make the biggest number of transfers⚽

```python

plt.figure(figsize=(8,5))
plt.hist(df['Age'], bins=10, log=False, color = 'green')
plt.title('Distribution of Football Players Age', fontsize=16)
plt.xlabel('Age of the football player')
plt.ylabel('Frequency')
plt.show()
```

#### ⚽As we can see players change their clubs many times⚽

```python
df["Name"].value_counts().sort_values(ascending=False)[0:96]
```

#### ⚽Let's check the clubs that have bought the biggest amount of players⚽

```python
ax= px.treemap(df,path=["Team_to"])
ax.show()
```

Most of the players were bought by Inter, Chelsea, Manchaster City

#### ⚽Let's check the clubs that have sold the biggest amount of players⚽

```python
ax1= px.treemap(df,path=["Team_from"])
ax1.show()
```

As we can see Inter has bought and sold the most amount of players, it can be told that the clubs that have the biggest amount of players bought also have sold the most.

#### ⚽Let's check on which position players change clubs the most⚽

```python
dx=df['Position'].value_counts()
ax2= dx.plot(kind="bar")
```

As we can see the Centre Forward players change clubs the most.

#### ⚽Let's check which player has maid the most money⚽

```python
highest_price = df.groupby('Name').sum()['Transfer_fee'].reset_index().sort_values(by='Transfer_fee', ascending=False)
display(highest_price)
```
