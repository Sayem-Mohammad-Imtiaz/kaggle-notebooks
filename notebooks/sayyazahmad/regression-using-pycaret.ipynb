# **PyCaret**

PyCaret is an open source, low-code machine learning library in Python that allows you to go from preparing your data to deploying your model within minutes in your choice of notebook environmen
https://pycaret.org/

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../input/real-estate-price-prediction/Real estate.csv')
```

```python
data.describe()
```

```python
data.drop(['No', 'X1 transaction date'], axis=1, inplace=True)
data.columns = ['Age', 'DistanceToMTR', 'NumStores', 'Latitude', 'Longitude', 'UnitPrice']
```

```python
plt.figure(figsize=(14, 6))
plt.hist(data['UnitPrice'])
```

```python
data = data[data['UnitPrice'] <= 70]
```

```python
plt.figure(figsize=(14, 6))
plt.hist(data['UnitPrice']);
```

***Model Setup***

```python
# Install PyCaret
!pip install pycaret
```

```python
from pycaret.regression import *
```

```python
exp_reg = setup(data, target='UnitPrice')
```

```python
compare_models()
```

**Random Forest performed pretty good, so we are going to pick Random Forest**

```python
model = create_model('rf')
```

**Model Tunning**

```python
tuned_model = tune_model(model)
```

**Model Visualization and Evaluation**

```python
plot_model(tuned_model)
```

```python
evaluate_model(tuned_model)
```

```python
interpret_model(tuned_model)
```

**Evaluation**

```python
predictions = predict_model(tuned_model)
```

```python
finalize_model(tuned_model)
```

**See how easy to create and fine tune model in PyCaret.**
