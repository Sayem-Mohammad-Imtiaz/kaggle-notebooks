```python
df1 = df.set_index(["treatment","gender"])
df1
# lets unstack it
```

```python
# level determines indexes
df1.unstack(level=0)
```

```python
df1.unstack(level=1)
```

```python
# change inner and outer level index position
df2 = df1.swaplevel(0,1)
df2
```
