<h2 style='border:3px solid DodgerBlue;'><center>After Feature Selection</center></h2>

```python
new_df = df.iloc[:,[0,1,2,3,4,5,6,9,10,11]]
```

```python
features = new_df.drop("Reached.on.Time_Y.N",axis=1) 
target   = new_df["Reached.on.Time_Y.N"]
```

```python
x_train,x_test,y_train,y_test = train_test_split(features,target,random_state=32,test_size=0.30)
```

```python
clf = DecisionTreeClassifier() 
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
print("Training score:",clf.score(x_train,y_train)*100) 
print("Testing score :",clf.score(x_test,y_test)*100)
```
