Basic understanding for column meanings

life ladder:The top of the ladder represents the best possible life for you and the bottom of the ladder represents the worst possible life for you
    
GDP per capita:GDP per capita stands for Gross Domestic Product (GDP) per capita (per person). It is derived from a straightforward division of total GDP by the population.   

Social support: Social support means having friends and other people, including family, to turn to in times of need or crisis 
   
Healthy life expectancy at birth: The average equivalent number of years of full health that a newborn could expect to live
    
Freedom to make life choices: Freedom of choice describes an individual's opportunity and autonomy to perform an action selected from at least two available options, unconstrained by external parties.
    
Generosity: the quality of kindness and generous.
    
Perceptions of corruption: perceived levels of public sector corruption, as determined by expert assessments and opinion surveys

```python
southern_asia = ["India","Bangladesh","Pakistan","Afghanistan","Sri Lanka","Maldives","Nepal","Bhutan"]
```

<h2 style='border:3px solid DodgerRed;background-color:#9d0aea;font-family:comic sans ms'><center>Southern Asian Countries Rank based on Life Ladder</center></h2>

```python
life = df1.groupby("Country name")["Life Ladder"].mean().reset_index() 
life["Index"] = [i for i in range(len(life))] 
life = life[life["Country name"].isin(southern_asia)]  
life.columns = ['Country name', 'Life Ladder', 'Rank'] 
life = life.loc[life["Life Ladder"].nlargest(10).index.tolist(),:] 
life["Rank"] = [i for i in range(1,len(life["Rank"])+1)] 
ff.create_table(life)
```

```python
southern = df1[df1["Country name"].isin(southern_asia)] 
plt.figure(figsize=(15,10))
px.line(southern,x="year",y="Life Ladder",color="Country name",template="plotly_dark")
```

<h2 style='border:3px solid DodgerRed;background-color:#9d0aea;font-family:comic sans ms'><center>Southern Asian Countries Rank based on Log GDP Per capita</center></h2>

```python
gdp = df1.groupby("Country name")["Log GDP per capita"].mean().reset_index() 
gdp["Index"] = [i for i in range(len(gdp))] 
gdp = gdp[gdp["Country name"].isin(southern_asia)]  
gdp.columns = ['Country name', 'Log GDP per capita', 'Rank'] 
gdp = gdp.sort_values(by="Log GDP per capita") 
gdp["Log GDP per capita"].nlargest(10).index 
gdp = gdp.loc[[90, 138, 15, 63, 113, 10, 103, 0],:] 
gdp["Rank"] = [i for i in range(1,len(gdp["Rank"])+1)] 
ff.create_table(gdp)
```

```python
southern = df1[df1["Country name"].isin(southern_asia)] 
plt.figure(figsize=(15,10))
px.line(southern,x="year",y="Log GDP per capita",color="Country name",template="plotly_dark")
```
