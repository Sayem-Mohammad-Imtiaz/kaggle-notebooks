## preprocessing 

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df["label"] = LE.fit_transform(df["label"])
df['label'].unique()
```

```python
df["label"].value_counts()
```

```python
df.corr().abs()['label'].sort_values(ascending = False)
```

## Split the Dataset Training 80 % and Tsting 20 %

```python
import sklearn.model_selection
train, test = sklearn.model_selection.train_test_split(df,stratify=df["label"], train_size=.80, random_state=1)
print(train.shape, test.shape)
```

<h1 style='background-color:#4C66E0; font-family:newtimeroman; font-size:190%; text-align:center; border-radius: 15px 50px;' >Deep_Autoviml</h1>


#### Build keras pipelines and models in a single line of code!

<img src="https://github.com/AutoViML/deep_autoviml/raw/master/logo.jpg" width="800px">

#### Link 
[Here](https://github.com/AutoViML/deep_autoviml)

```python
target = 'label'
```

```python
keras_model_type =  "fast1" ## always try "fast" first, then "fast2", "auto", etc.
### always set early_stopping to True first and then change it to False
#### You always need 15 max_trials to get something decent #####
#### always set tuner to "storm" and then "optuna". 

project_name = "Gender Recognition by Voice Identify a voice as male or female"
model_options = {'char_limit':50, 'cat_feat_cross_flag':False,
                 'max_trials': 10, "tuner": "storm"}
keras_options = {"patience":10, 'class_weight': True, 'early_stopping': True, 
                 'lr_scheduler': '', "optimizer": 'RMS'}
```

```python
model, cat_vocab_dict = deepauto.fit(train, target, keras_model_type=keras_model_type,
        project_name=project_name, keras_options=keras_options,  
        model_options=model_options, save_model_flag=True, use_my_model='',
        model_use_case='', verbose=1)
```

```python
predictions = deepauto.predict(model, project_name, test_dataset=test,
                                 keras_model_type=keras_model_type, 
                                 cat_vocab_dict=cat_vocab_dict)
```

```python
y_test = test[target].values
y_test
```

```python
y_preds = predictions[-1]
y_preds
```
