# EDA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../input/netflix-shows/netflix_titles.csv')
```

data info

```python
# dataset check
df.head(5)
```

```python
# data type & missing value check
df.info()
```

```python
# rows and columns(7787 records and 12 columns)
df.shape
```

```python
# sum of null values
df.isnull().sum()
```

```python
# missing values ratio
df.isnull().sum()/len(df)*100
```

```python
# since the missing value ratio of director is high, I will drop it

df.drop('director', axis = 1, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.date_added.fillna('unknown', inplace=True)
```

```python
# Data preprocessing : date_added -> year_added
df['year_added'] = df.date_added.apply(lambda x : str(x).split(',')[-1])
```

```python
# Replace missing values with maximum
df.year_added.value_counts()

df['year_added'] = df.year_added.apply(lambda x : str(x).replace('unknown', '2019'))
```

```python
# removing extra space
df['year_added'] = df.year_added.apply(lambda x : str(x).replace(' ', ''))
```

```python
df.year_added.value_counts()
```

```python
df.isnull().sum()
```

```python
#displaying the most occurred country
df.country.value_counts()
```

```python
#filling null values of country with most occuring country
df.country.fillna("United_states",inplace= True)
```

```python
#displaying the most occurred rating
df.rating.value_counts()
```

```python
#filling null values of rating with most occuring rating
df.rating.fillna("TV-MA",inplace= True)
```

# Visualization

```python
# displaying all the countries with most contribution to netflix content
x= df.country.value_counts().head(20)
plt.figure(figsize=(10,10))
sns.set_style("whitegrid")   #set style to whitegrid
ax= sns.barplot(x.values,x.index)   #using bar for visualization
ax.set_xlabel("Number of content")
ax.set_ylabel("Number of country")
```

```python
# distribution of type of content on netflix 
plt.figure(figsize=(10,10))

x= df.type.value_counts()
sns.countplot(x= "type",data= df)
```

```python
plt.figure(figsize=(15,5))
sns.histplot(df.release_year[df.type=='Movie'],color='r')
sns.histplot(df.release_year[df.type=='TV Show'],color='y')
plt.legend(['Movie','TV Show'])
plt.show()
```

```python
# content distribution based on rating
plt.figure(figsize=(12,10))
sns.set(style="darkgrid")
ax = sns.countplot(x="rating", data=df,order=df['rating'].value_counts().index[0:20])
```

```python
# content distribution based on rating : South Korea
plt.figure(figsize=(12,10))
sns.set(style="darkgrid")
ax = sns.countplot(x="rating", data=df,order=df.rating[df.country=='South Korea'].value_counts().index[0:20])
```

```python
x= df.rating.value_counts()
plt.figure(figsize=(10,10))

labels= list(x.index)
plt.pie(x.values,labels= labels,autopct= "%1.1f%%")
plt.show()
```
