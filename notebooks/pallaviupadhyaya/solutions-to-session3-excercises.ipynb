```python
# Total number of non-smokers from northeast region
# Selecting regions corresponding to smokers as an array
x = np.array(insurance[insurance.smoker == 'no'].region)
print('number of non smokers in all regions',len(x))
print('number of non smokers in northeast')
print ((x == 'northeast').sum())
```

```python

# Average expenses of men from southeast who are smokers
y = insurance.loc[(insurance['region']=='southeast') & (insurance['sex'] == 'male') & (insurance['smoker']=='yes')]
print("Average expenses", y.expenses.mean())
print('number of men who are smokers in southeast',len(y))

```

```python
# Total men with bmi > 28 and women with bmi> 26
z = insurance.loc[(((insurance['bmi']> 28) & (insurance['sex'] == 'male')) | ((insurance['sex']=='female') & (insurance['bmi']> 26)))]
print("total", len(z))
```
