### Step 3 : Explore the Data
Here I am changing the column names of the Data Frame from the long Descriptive names to shorter ones. This is not a required step but I am doing it just for convenience. 

It helps me to visualize the data in a more compact manner.

```python
df.columns = ['name', 'bar_name', 'ref', 'year', 'coco_percent', 'location', 'rating', 'type', 'bean origin']
df.head()
```

As you have noticed above, "coco_percent" field is a string because it has the "%" suffix. 

We need to change it to number. For this below I am using the "apply" method on Pandas DataFrame using "lambda" function.

All I am doing is reading each record in the "coco_percent" column and stripping the "%" character and the converting the remaining string to number which should be "float" and not "integer" which should be self explanatory.

```python
df['coco_percent'] = df['coco_percent'].apply(lambda x: float(x.strip('%')))
df.head()
```

I will now start explore the features one by one.

Here I am looking at the frequencey count and trying to make some plots.

```python
df['name'].value_counts(dropna = False)
```

Below is the plot for Top 50 - Chocolate makers based on count.

```python
df['name'].value_counts(dropna = False)[:50].plot(kind = 'bar', figsize = (15, 10))
```

Below is the same plot but for the last 50 or tail . 

```python
df['name'].value_counts(dropna = False)[-50:].plot(kind = 'bar', figsize = (15, 10))
```

Next I will move to the bar name.  Looks its more like place of origin than bar name which I thought initially it was.

```python
df['bar_name'].value_counts(dropna = False)
```

**Top 20 place of origin.**

```python
df['bar_name'].value_counts(dropna = False)[:25].plot(kind = 'bar', figsize = (20, 10))
```
