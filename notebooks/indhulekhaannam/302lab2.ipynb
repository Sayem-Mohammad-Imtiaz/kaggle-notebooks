```python
data3 = {'Cp':[16, 15, 31],
        'Cr':[10, 15, 25],
        'N' : [9, 35, 44],
        'Total': [35, 65, 100]}
 
# Creates pandas DataFrame.
df3 = pd.DataFrame(data3, index =['Boys',
                                'Girls',
                                'Total'])
df3
```

## Probability that randomly selected student is a boy = Number of boys/ Total number of students

```python
prob_boy1 = df1['Total']['Boys']/df1['Total']['Total']
print("In Distribution 1 - ", prob_boy1)
prob_boy2 = df2['Total']['Boys']/df2['Total']['Total']
print("In Distribution 2 - ", prob_boy2)
prob_boy3 = df3['Total']['Boys']/df3['Total']['Total']
print("In Distribution 3 - ", prob_boy3)
```

## Probability that randomly selected student is a boy and covid positive

```python
prob_covid_boy1 = df1['Cp']['Boys'] / df1['Total']['Total']
print("In Distribution 1 - ", prob_covid_boy1)
prob_covid_boy2 = df2['Cp']['Boys'] / df2['Total']['Total']
print("In Distribution 2 - ", prob_covid_boy2)
prob_covid_boy3 = df3['Cp']['Boys'] / df3['Total']['Total']
print("In Distribution 3 - ", prob_covid_boy3)
```

## Probability that randomly selected student is either covid positive or recovered

```python
prob_covd_reco1 = (df1['Cp']['Total'] +  df1['Cr']['Total'])/ df1['Total']['Total']
print("In Distribution 1 - ", prob_covd_reco1)
prob_covd_reco2 = (df2['Cp']['Total'] +  df2['Cr']['Total'])/ df2['Total']['Total']
print("In Distribution 2 - ", prob_covd_reco2)
prob_covd_reco3 = (df3['Cp']['Total'] +  df3['Cr']['Total'])/ df3['Total']['Total']
print("In Distribution 3 - ", prob_covd_reco3)
```

## Probability that randomly selected student is a girl who is covid positive or recovered

```python
prob_covd_reco_girl1 = (df1['Cp']['Girls'] +  df1['Cr']['Girls'])/ df1['Total']['Total']
print("In Distribution 1 - ", prob_covd_reco_girl1)
prob_covd_reco_girl2 = (df2['Cp']['Girls'] +  df2['Cr']['Girls'])/ df2['Total']['Total']
print("In Distribution 2 - ", prob_covd_reco_girl2)
prob_covd_reco_girl3 = (df3['Cp']['Girls'] +  df3['Cr']['Girls'])/ df3['Total']['Total']
print("In Distribution 3 - ", prob_covd_reco_girl3)
```

```python
# a function to calculate expectation
def calc_Expectation(a):
    sum = 0
    sum += a[0] * a[0]/100 + a[1]*a[1]/100  
    return float(sum)
```

```python
# modifying the dataframe
df1_ = df1.drop('Total')
df2_ = df2.drop('Total')
df3_ = df3.drop('Total')
```

## Expectation of current covid +ve

```python
arr_Cp1 = df1_['Cp'].to_numpy()
expect_covid_positive_curr1 = calc_Expectation(arr_Cp1)
print("Current covid +ve expectation in Distribution 1 is", expect_covid_positive_curr1)
arr_Cp2 = df2_['Cp'].to_numpy()
expect_covid_positive_curr2 = calc_Expectation(arr_Cp2)
print("Current covid +ve expectation in Distribution 2 is", expect_covid_positive_curr2)
arr_Cp3 = df3_['Cp'].to_numpy()
expect_covid_positive_curr3 = calc_Expectation(arr_Cp3)
print("Current covid +ve expectation in Distribution 3 is", expect_covid_positive_curr3)
```

## Expectation of recovered from covid

```python
arr_Cr1 = df1_['Cr'].to_numpy()
expect_covid_positive_recovered1 = calc_Expectation(arr_Cr1)
print("Recovered covid +ve expectation in Distribution 1 is", expect_covid_positive_recovered1)
arr_Cr2 = df2_['Cr'].to_numpy()
expect_covid_positive_recovered2 = calc_Expectation(arr_Cr2)
print("Recovered covid +ve expectation in Distribution 2 is", expect_covid_positive_recovered2)
arr_Cr3 = df3_['Cr'].to_numpy()
expect_covid_positive_recovered3 = calc_Expectation(arr_Cr3)
print("Recovered covid +ve expectation in Distribution 3 is", expect_covid_positive_recovered3)
```

## Expectation of "Never affected from covid"

```python
arr_N1 = df1_['N'].to_numpy()
expect_covid_not_affected1 = calc_Expectation(arr_N1)
print("Never affected by covid expectation in Distribution 1 is", expect_covid_not_affected1)
arr_N2 = df2_['N'].to_numpy()
expect_covid_not_affected2 =calc_Expectation(arr_N2)
print("Never affected by covid expectation in Distribution 2 is", expect_covid_not_affected2)
arr_N3 = df3_['N'].to_numpy()
expect_covid_not_affected3 = calc_Expectation(arr_N3)
print("Never affected by covid expectation in Distribution 3 is", expect_covid_not_affected3)
```
