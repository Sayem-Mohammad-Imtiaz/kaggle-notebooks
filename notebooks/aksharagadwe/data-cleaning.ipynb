```python
genre = []
```

```python
def genre_dict(string):
    x = string.split(", ")
    for genre_type in x:
        if genre_type not in genre:
            genre.append(genre_type)
```

```python
for values in data.listed_in:
    genre_dict(values)
```

```python
for i in range(data.shape[0]):
    temp = data.loc[i,"listed_in"]
    temp = temp.split(", ")
    for value in temp:
        data.at[i,value]=1
```
