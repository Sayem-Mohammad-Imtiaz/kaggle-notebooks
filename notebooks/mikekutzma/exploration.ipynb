We do some basic exploratory analysis on the homicide reports dataset.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

df = pd.read_csv("../input/database.csv",low_memory=False)

df.info()
```

```python
vic_col = "Victim Race"
perp_col = "Perpetrator Race"
vic_races = sorted(df[vic_col].unique())
perp_races = sorted(df[perp_col].unique())
crimes_against_by = df.groupby([vic_col,perp_col]).size()
fig,ax = plt.subplots(figsize=(10,5))
pos =  list(range(len(vic_races)))
width = 0.15

for race,ind in zip(vic_races,range(0,len(vic_races))):
    plt.bar([p+(ind*width) for p in pos],
           np.log10(crimes_against_by[race].values),
           width,
           alpha=0.7,
           label=race)

ax.set_xticks([p + (width*(len(perp_races)-1)/2.0) for p in pos])
ax.set_xticklabels(vic_races)
ax.set_xlabel("Victim Race")
ax.set_ylabel("Number of murders (log scale)")
plt.legend(perp_races)
```

Each grouping on the x-axis represents a victim race, with each column within the grouping representing a perpetrator race. The number of murders within each victim/perp pair is represented by the height of the bar on the y-axis, plotted on a logarithmic scale. Due to the the large discrepancy in number of murders between each pair, a log scale more accurately displays the information. The log scale allows for each unit in the y-axis to represent a ten-fold increase in murders, begining with '0' representing a single murder.
