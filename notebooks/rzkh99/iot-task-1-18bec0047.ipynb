```python
correlation = vaccines.corr()
```

```python
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns, annot=True)
```

```python
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns, annot=False)
```

```python
sns.pairplot(vaccines)
```

```python
sns.relplot(x = 'daily_vaccinations', y = 'people_vaccinated', hue = 'vaccines', data = vaccines)
```
