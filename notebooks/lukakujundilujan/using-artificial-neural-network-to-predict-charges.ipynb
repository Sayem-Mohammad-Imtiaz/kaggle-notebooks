```python
#as we concluded , we are droping region data, because i think it is totaly irrelevant for our model, 

df2=df2.drop("region", axis=1)
```

```python
#if we want to make our model we need to convert everything into numbers. we are makeing new categories from sex, childern and smoker variable
dummies = pd.get_dummies(df2[['sex', 'children','smoker']],drop_first=True)
df2 = df2.drop(['sex', 'children','smoker'],axis=1)
df2 = pd.concat([df2,dummies],axis=1)
df2.T
```

```python
#time to import our model
from sklearn.model_selection import train_test_split
```

```python
#prepare our model. 
X =df2.drop("charges", axis=1).values
y =df2["charges"].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
#to use tensorflow keras models we need to scale variables
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
X_train.shape
```

```python
#i am not an expert in this. I watched one turtorial. Here is Sequental model. 
#We have 4 "hidden layers" with 7 neurons , as our X_train shape have 7 input variables
model =Sequential()
model.add(Dense(7, activation="relu"))
model.add(Dense(7, activation="relu"))
model.add(Dense(7, activation="relu"))
model.add(Dense(7, activation="relu"))
#aparently optimizer adam is best one.there is more on wikipedia for "relu" and "adam". i dont know very much into details to be honest
model.add(Dense(1))
model.compile(optimizer="adam", loss="mse",metrics=[])
```

```python
model.fit(x=X_train, y=y_train, validation_data=(X_test, y_test), batch_size=128, epochs=400)
```

```python
#how or model was performing
losses =pd.DataFrame(model.history.history)
losses.plot()
```

```python
#lets check how far away we are from predicted values
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score, r2_score
```

```python
predictions =model.predict(X_test)
```

```python
#on average our model is missing about 4000$. one month ago i used RandomForestRegressor from sklearn and i was off about 2500
#that proves that RandomForestRegressor could be better for simple lineral models. 
mean_absolute_error(y_test, predictions)
```
