Putting the new values in a new colunm "Reason"

```python
df['Reason'] = df["title"].apply(lambda title: title.split(':')[0])
```

```python
df['Reason']
```

** What is the most common Reason for a 911 call based off of this new column? **

```python
df['Reason'].value_counts()
```

** Now use seaborn to create a countplot of 911 calls by Reason. 

```python
sns.countplot(x='Reason',data=df,palette='Paired')
```

as we can see timeStamp column value is a str 

```python
df.info()
```

```python
type(df['timeStamp'].iloc[0])
```

converting timeStamp to datetime

```python
df['timeStamp']= pd.to_datetime(df['timeStamp'])
```

```python
type(df['timeStamp'].iloc[0])
```

splitting timeStamp column in three new columns: hour;month;day of week.

```python
df['hour']= df['timeStamp'].apply(lambda time:time.hour)
df['month']= df['timeStamp'].apply(lambda time:time.month)
df['day of week']= df['timeStamp'].apply(lambda time:time.day)
```

```python
df.head()
```

```python
giving names to the day of week
```

```python
dmp = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['day of week'] = df ['day of week'].map(dmp)
```

```python
df.head()
```

Visualization of Reasons according to day of week and month 

```python
#Reason according to day of week
sns.countplot(x='day of week',data=df,hue='Reason',palette='viridis')
#organizing the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
#Reason according to month
sns.countplot(x='month',data=df,hue='Reason',palette='Paired')
#organizing the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

 amount of calls by months

```python
bymonth = df.groupby('month').count()
```

```python
bymonth['lat'].plot()
```

```python
sns.lmplot(x='month',y='twp',data=bymonth.reset_index())
```

Creating a new comlumn date to from timeStamp column

```python
df['date']=df['timeStamp'].apply(lambda t: t.date())
```

TWP(Township) 911 call spike dates

```python
df.groupby('date').count()['twp'].plot()
plt.tight_layout()
```

TWP(Township) 911 call spike for traffic

```python
df[df['Reason']=='Traffic'].groupby('date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
```

TWP(Township) 911 call spike for fire

```python
df[df['Reason']=='Fire'].groupby('date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
```

TWP(Township) 911 call spike for EMS

```python
df[df['Reason']=='EMS'].groupby('date').count()['twp'].plot()
plt.title('EMS')
plt.tight_layout()
```
