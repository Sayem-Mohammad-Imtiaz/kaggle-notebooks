# Preparing Data for Fitting 

### If a wine has quality value over 5 its is a good wine and if value is below 5 its bad wine

```python
# So changing quality value to True or False Based on if a wine has quality value over 5 
#(if quality > 5 its is a good wine or else its False)

y=data_import['quality']>5
y
```

```python
# Removing Target Variable i.e Quality

X=data_import.iloc[:,:-1]

X
```

# Splitting data into Training data and Testing data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)
```

```python
X_train.shape
```

```python
X_test.shape
```

## What is Standard Scaler ?

### StandardScaler standardizes a feature by subtracting the mean and then scaling to unit variance. Unit variance means dividing all the values by the standard deviation. 
### StandardScaler makes the mean of the distribution 0. About 68% of the values will lie be between -1 and 1.

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

scaler_X_train=scaler.fit_transform(X_train)

scaler_X_train
```

```python
scaler_X_test=scaler.transform(X_test)

scaler_X_test
```

```python
scaler_X_test
```

```python
X_train.shape
```

# Building SVM Classifier

```python
from sklearn.svm import SVC

svc_clf = SVC(C=1.0, kernel='rbf', degree=3, gamma='auto', coef0=0.0, shrinking=True, probability=False, 
              tol=0.001, cache_size=200, class_weight=None, verbose=False, max_iter=-1, decision_function_shape='ovr', 
              break_ties=False,random_state=None)

svc_clf.fit(scaler_X_train,y_train)
```

```python
svc_clf_predictions=svc_clf.predict(scaler_X_test)
```

```python
svc_clf_predictions
```

```python
c=confusion_matrix(y_test,svc_clf_predictions)
a=accuracy_score(y_test,svc_clf_predictions)
p=precision_score(y_test,svc_clf_predictions)
r=recall_score(y_test,svc_clf_predictions)
```

```python
print('Confusion Matrix:\n',c)
```

```python
print('Accuracy:',a*100)
```

```python
print('Precision:',p*100)
```

```python
print('Recall:',r*100)
```

### Visualizing Confusion matrix using seaborn 

```python
fig, ax = plt.subplots(figsize=(7.5, 7.5))
ax.matshow(c, cmap=plt.cm.Blues, alpha=0.3)
for i in range(c.shape[0]):
    for j in range(c.shape[1]):
        ax.text(x=j, y=i,s=c[i, j], va='center', ha='center', size='xx-large')
plt.xlabel('Predictions', fontsize=18)
plt.ylabel('Actuals', fontsize=18)
plt.title('Confusion Matrix', fontsize=18)
plt.show()
```
