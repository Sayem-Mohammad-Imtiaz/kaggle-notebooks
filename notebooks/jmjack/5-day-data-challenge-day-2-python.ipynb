Importing regex library to clean the data

```python
import pandas as pd
import matplotlib.pyplot as plt
import re
```

Reading and cleaning data from N/A

```python
data = pd.read_csv("../input/salaries-by-college-type.csv")
data.dropna(inplace=True)
data.reset_index(inplace=True, drop=True)
data.head()
```

Changing strings to floats by removing "$ . ," using regex, then casting to float. By doing so we can perform numerical operations on data and plot histograms.

```python
for column in range(2,8):
    data.iloc[:, column] = [float(re.sub('[$,.]', '', i)[:-2]) for i in data.iloc[:, column]]
```

Checking if the data was properly cleaned

```python
data
```

```python
plt.hist(data['Starting Median Salary'])
plt.title("Histogram of Starting Median Salary")
```

```python
plt.hist(data['Mid-Career Median Salary'], color='navy')
plt.title("Histogram of Mid-Career Median Salary")
```
