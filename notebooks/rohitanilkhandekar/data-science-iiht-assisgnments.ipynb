1) Linear Equation Sums

4x  + 3y = 20;
-5x + 9y = 26

```python
import numpy as np
m_list=[[4,3],[-5,9]]
A=np.array(m_list)
print('Matrix A:')
print(A)

n_list=[[20],[26]]
B=np.array(n_list)
print('Matrix B:')
print(B)

X=np.linalg.inv(A).dot(B)
print('x and y are:')
print(X)
```

4x + 3y + 2z = 25; 
-2x + 2y + 3z = -10; 
3x -5y + 2z = -4

```python
import numpy as np
A= np.array([[4,3,2],[-2,2,3],[3,-5,2]])
print('Matrix A:')
print(A)
B=np.array([25,-10,-4])
print('Matrix B:')
print(B)
X=np.linalg.inv(A).dot(B)
print('Values of x,y and z are:')
print(X)
```
