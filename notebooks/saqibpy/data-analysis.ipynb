```python
import pandas as pd

df = pd.read_csv('../input/pakistanintellectualcapitalcs/Pakistan Intellectual Capital - Computer Science - Ver 1.csv',encoding='latin1')
df.set_index('S#',inplace=True)
df.head()
```

```python
df['University Currently Teaching'].value_counts().plot(kind='bar', title ="Universty / College", figsize=(24, 4), legend=True, fontsize=12)
```

```python
df['Terminal Degree'] = df['Terminal Degree'].str.strip()\
                .replace(regex='(BS|BSCS)'           ,value='BS')\
                .replace(regex='(MCS|MSCS|M.S)'      ,value='MS')\
                .replace(regex='(MCS|MSCS|M.S|M.E)'  ,value='MS')\
                .replace(regex='(M.Phil|Mphil|MPhil)',value='MPhil')\
                .replace(regex='(Phd|Ph.D)'          ,value='PhD')\
                .replace('PhD(Scholar)','PhD').replace('PhD (Scholar)','PhD')
                            

df['Terminal Degree'].value_counts()



# df['Terminal Degree'].value_counts().plot(kind='pie', title ="Qualification", figsize=(8, 8), fontsize=12)
```

```python
df['Province University Located'].value_counts().plot(kind='pie', title ="Locality", figsize=(8, 8), fontsize=12)
```
