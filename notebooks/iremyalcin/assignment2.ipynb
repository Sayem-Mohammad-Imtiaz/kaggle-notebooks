## Basic Data Analysis

<a id="7"> </a><br>
### 15 Most Common Author

```python
authorCount = Counter(data['Author'])
mostAuthor = authorCount.most_common(15)

x,y = zip(*mostAuthor)
x,y = list(x),list(y)

plt.figure(figsize=(30,10))
sns.barplot(x=x,y=y,palette=sns.cubehelix_palette(len(x)))
plt.show()
```

<a id="8"> </a><br>
### Author & Price

```python
authorPrice = []
dataMostAuthor = pd.DataFrame(mostAuthor,columns=['author','count'])

for i in dataMostAuthor['author']:
    x = data[data['Author']==i]
    price = sum(x['Price']) / len(x)
    authorPrice.append(price)
    
dataMostAuthor['price'] = authorPrice

plt.figure(figsize=(10,5))
sns.barplot(x=dataMostAuthor['author'],y=dataMostAuthor['price'])
plt.xticks(rotation=90)
plt.xlabel('Author')
plt.ylabel('Price')
plt.title('15 Most Common Authorses Average Price')
plt.show()

dataMostAuthor[["author","price"]].groupby(["author"], as_index = False).mean().sort_values(by="price",ascending = False)

mostAuthorList=list(dataMostAuthor['author'])
```

<a id="9"> </a><br>
### Author & Year

```python
year = []

for i in dataMostAuthor['author']:
    x = data[data['Author']==i]
    yearAve = sum(x['Year']) / len(x)
    year.append(yearAve)
    

plt.figure(figsize=(10,5))
sns.barplot(x=dataMostAuthor['author'],y=year)
plt.xticks(rotation=90)
plt.xlabel('Author')
plt.ylabel('Price')
plt.title('15 Most Common Authorses Years')
plt.ylim(2005,2021)
plt.show()

dataMostAuthor['year'] = year
dataMostAuthor[["author","year"]].groupby(["author"], as_index = False).mean().sort_values(by="year",ascending = False)
```

<a id="10"> </a><br>
### Author & User Ratings

```python
rate = []

for i in dataMostAuthor['author']:
    x = data[data['Author']==i]
    rateAve = sum(x['User Rating']) / len(x)
    rate.append(rateAve)
    

plt.figure(figsize=(10,5))
sns.barplot(x=dataMostAuthor['author'],y=rate)
plt.xticks(rotation=90)
plt.xlabel('Author')
plt.ylabel('Rate')
plt.title('15 Most Common Authorses Rate')
plt.ylim(3.8,5)
plt.show()

dataMostAuthor['rate'] = rate
dataMostAuthor[["author","rate"]].groupby(["author"], as_index = False).mean().sort_values(by="rate",ascending = False)
```

## Outlier Detections & Missing Value

```python
def detectOutliers(data,features):
    outlierIndices=[]
    
    for f in features:
        
        Q1 = np.percentile(data[f],25)
        Q3 = np.percentile(data[f],75)
        
        IQR = Q3-Q1
        outlierStep = IQR * 1.5
        
        outlierListCol = data[(data[f]<Q1 - outlierStep) | (data[f]>Q3 + outlierStep) ].index
        outlierIndices.extend(outlierListCol)
        
    outlierIndices = Counter(outlierIndices)
    multipleOutliers = list(i for i, v in outlierIndices.items() if v>2)
    return multipleOutliers
```

```python
data.loc[detectOutliers(data,["Year","User Rating","Price"])]
```

```python
yearList = list(data['Year'].unique())
dataYearGenre = pd.DataFrame(yearList,columns=['year'])

finc = []
nonF = []

for i in yearList:
     x = data[data['Year'] == i]
     finc.append(sum(x['Genre'] == 'Fiction'))
     nonF.append(sum(x['Genre'] == 'Non Fiction'))
        
dataYearGenre['finC'] = finc
dataYearGenre['nonfC'] = nonF

f,ax = plt.subplots(figsize = (9,9))
sns.barplot(x='year',y='finC',data=dataYearGenre,alpha = 0.5,label='Fiction',color='red')
sns.barplot(x='year',y='nonfC',data=dataYearGenre,alpha = 0.7,label='Non Fiction',color='red')

ax.legend(loc='lower right',frameon = True)
ax.set(xlabel='Fiction and Non Fiction of Years', ylabel='States',title = "Percentage of State's Population According to Races ")
plt.show()
```
