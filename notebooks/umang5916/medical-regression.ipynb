```python
#Defining X, y
y = df.iloc[:, 3].values
df = df.drop(['charges'], axis = 1)
X = df.iloc[:, :].values
```

```python
#Splitting Dataset for training and testing the model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0) 
```

```python
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of Y_train:", y_train.shape)
print("Shape of Y_test:", y_test.shape)
```

**Regression Model Applied:**
1. MultiLinear Regression
2. SVR
3. Decision Tree Regression
4. Random Forest Regression
5. XGB Regression

```python
#Multi-Linear Regression
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
lr_pred = lr.predict(X_test) 
```

```python
#SVR
from sklearn.svm import SVR
svr = SVR(kernel = 'rbf')
svr.fit(X_train, y_train)
svr_pred = svr.predict(X_test)
```

```python
#Decision Tree Regression
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor(random_state = 0)
dt.fit(X_train, y_train)
dt_pred = dt.predict(X_test) 
```

```python
#Random Forest Regression
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 200, random_state = 0)
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test) 
```

```python
#XGB Regression
from xgboost import XGBRegressor
xgb = XGBRegressor()
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test) 
```

```python
print("RMSE valuse for MultiLinear Regression:", sqrt(mse(y_test, lr_pred)))
print("RMSE valuse for SVR:", sqrt(mse(y_test, svr_pred)))
print("RMSE valuse for Decision Tree Regression:", sqrt(mse(y_test, dt_pred)))
print("RMSE valuse for Random Forest Regression:", sqrt(mse(y_test, rf_pred)))
print("RMSE valuse for XGB Regression:", sqrt(mse(y_test, xgb_pred)))
```

**From the above applied regression models we can see clearly that Random Forest Regression gives us the lowest RMSE thus being the most accurate model for this Dataset.
Thus to finally predict a value we will use Random Forest Regression**
