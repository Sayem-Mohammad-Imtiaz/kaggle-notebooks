Aşağıda tablo hakkındaki genel bilgilere baktık.


```python
data=pd.read_csv("../input/bitcoin-historical-data/bitstampUSD_1-min_data_2012-01-01_to_2020-04-22.csv")
data.head()
data.tail(5)
data.columns
data.info()
```

Aşağıda tablo sütunları arasındaki oransallık ilişkilerine baktık.

```python
f,g=plt.subplots(figsize=(15,15))
sns.heatmap(data.corr(),annot=True, fmt='.1f',ax=g)
plt.show()
```

```python
x=np.min(data["Close"])
data[data["Close"]==x].Open  #en düşük kapanışta bitcoin açılışı ne kadardı?

data2=data["Close"]-data["Open"] #açılış kapanış farkı
print("bir günde en çok ne kadar kazandırdı?",data2.max())
print("bir günde en çok ne kadar kaybettirdi?",-data2.min())
data2.plot(kind="line",color="green",label="Açılış-Kapanış Farkı",grid=True,linestyle="-",figsize=(15,15))
data.High.plot(kind="line",color="red",label="max değer",grid=True,linestyle="--",figsize=(15,15))
data.Low.plot(kind="line",color="blue",label="min değer",grid=True,linestyle=":",alpha=0.5,figsize=(15,15))
plt.xlabel("Time")
plt.ylabel("Value")
plt.legend()
plt.show()


```

Yukarıdaki grafiğe göre bitcoin bir defa çok fazla kaybettirirken, bir defa da çok fazla kazandırmıştır.Geri kalan zamanlarda ani sıçrama görülmemektedir.gün içindeki max ve min değerlerinde bazen ani değişimler görülse de genellikle süprizlere az rastlanıyor.

Aşağıdaki tabloda da görüleceği gibi gün içinde max ve min değerleri genel itibari ile birbirine yakın değerlerdir.Sadece nadir zamanlarda ani değişimler görülmektedir. Bu açıklamayı yukarıdaki grafik de destekler.
