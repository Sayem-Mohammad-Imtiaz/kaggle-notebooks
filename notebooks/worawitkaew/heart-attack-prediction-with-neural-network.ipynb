```python
import numpy as np 
import pandas as pd 

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data  = pd.read_csv(dirname + '/heart.csv')
```

```python
import copy
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def nonlin(x, deriv=False):
    if deriv == True:
        return x*(1-x)
    return 1/(1+np.exp(-x))

class neural_network:
    def __init__(self, x, y,numberofnode):
        
        self.numberofnode = numberofnode
        self.num_len_of_x = x
        self.num_len_of_y = y
        self.syn0 = 2 + np.random.random((self.num_len_of_x ,self.numberofnode)) - 1
        self.syn1 = 2 + np.random.random((self.numberofnode ,self.num_len_of_y)) - 1
       
        
    def predict_proba(self,x):
        
        self.l0 = x
       
        self.l1 = nonlin(np.dot(self.l0,self.syn0))
        self.l2 = nonlin(np.dot(self.l1,self.syn1))
        
        return copy.copy(self.l2)
        
    def backward(self,y):  
        accelerator = 0.04
        
        l2_error = y - self.l2
        
        l2_delta = l2_error*nonlin(self.l2,deriv=True) * accelerator
        
        l1_error = l2_delta.dot(self.syn1.T)
        
        l1_delta = l1_error * nonlin(self.l1,deriv=True) * accelerator
        
        self.syn1 += self.l1.T.dot(l2_delta)
        self.syn0 += self.l0.T.dot(l1_delta)
        
        self.l1 = nonlin(np.dot(self.l0,self.syn0))
        self.l2 = nonlin(np.dot(self.l1,self.syn1))
```

```python
y = data[["output"]]
x = data.drop(["output"],axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
y = scaler.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
y_test = np.array(y_test)
y_train = np.array(y_train)
```

```python
model = neural_network(13,1,5)
```

```python
l2 = model.predict_proba(X_train)
for i in range(20000):
    model.backward(y_train)
```

```python
from sklearn.metrics import accuracy_score
result = model.predict_proba(X_test)
result = np.round(result)
accuracy = accuracy_score(y_test, result)
print("Accuracy : ",accuracy)
```
