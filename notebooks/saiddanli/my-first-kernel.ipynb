    Datamızın daha iyi anlaşılması için sütunların neyi ifade ettiğini açıklayalım:
1. age 
2. sex (1=erkek, 0=kadın)
3. cp (chest pain:göğüs ağrısı, değerler:0(ağrı yok)-1-2-3(çok şiddetli ağrı))
4. trestbps (dinlenme anındaki kan basıncı)
5. chol	(kolestrol mg/dl)
6. fbs (açlık kan şekeri, değerler:0(<120)-1(>120))
7. thalach (max kalp atım hızı)
8. target (kalp hastalığı, değerler:0(sağlıklı)-1(hasta))

```python
data.corr()
```

```python
f,ax = plt.subplots(figsize=(12,12))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt=".1f", ax=ax)
plt.show()
```

    Isı haritasında korelasyonları incelediğimiz zaman, yaş(age) ile max kalp atım hızı(thalach) arasında negatif korelasyon olduğu açıkca görülmektedir. Yani daha küçük yaşlarda max kalp atım hızı yüksekken, yaşlandıkça max kalp atım hızı azalmaktadır.
    Aşağıda yer alan scatter plot da bize bu negatif korelasyonu açıkca göstermektedir.

```python
data.plot(kind="scatter", x="age",y="thalach",color="red", alpha=.5)
plt.xlabel("age")
plt.ylabel("max_heart_rate")
plt.show()
```
