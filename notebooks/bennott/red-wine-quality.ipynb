# Red Wine Quality Prediction Using Simple Random Forest Regression
The methods that I used for prediction is fairly simple and not that accurate, since I'm using the method taught from the Kaggle micro course.

I used Random Forest Regression for the model and Cross-validation technique to determine the accuracy based on the MAE(Mean Absolute Error)

```python
import pandas as pd

# importing data
data = pd.read_csv('../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
data
```

```python
# searching for any missing values
na_count = pd.Series([data[col].isna().sum() for col in data.columns], index=data.columns, name='NA Count')
na_count
```

```python
# feature selection
y = data['quality']
X = data.drop(['quality'],axis=1)
print(y)
X
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# scoring function
def score(n):
    model = RandomForestRegressor(n_estimators=n, random_state=0)
    scores = -1 * cross_val_score(model,X,y,cv=5,scoring='neg_mean_absolute_error')
    
    return scores.mean()

# scores
for n in range(100,1001,100):
    print("n: {}, MAE: {}".format(n,score(n)))
```
