# Scaling Data

```python
X = df.drop('quality',axis=1)
y = df['quality']
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
```

```python
X.head()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=6)
model.fit(X)
```

```python
k_means_pred = model.predict(X)
k_means_pred
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
reduce_X = pd.DataFrame(pca.fit_transform(X),columns=['PCA1','PCA2'])
reduce_X
```

```python
# find the cluster center
reduce_centers = pca.transform(model.cluster_centers_)
reduce_centers
```

```python
reduce_X['Cluster'] = k_means_pred
```

```python
reduce_X
```

```python
plt.figure(figsize=(12,6))
plt.scatter(reduce_X[reduce_X['Cluster'] == 0].loc[:,'PCA1'],reduce_X[reduce_X['Cluster'] == 0].loc[:,'PCA2'],color='red',label='cluster1')
plt.scatter(reduce_X[reduce_X['Cluster'] == 1].loc[:,'PCA1'],reduce_X[reduce_X['Cluster'] == 1].loc[:,'PCA2'],color='green',label='cluster2')
plt.scatter(reduce_X[reduce_X['Cluster'] == 2].loc[:,'PCA1'],reduce_X[reduce_X['Cluster'] == 2].loc[:,'PCA2'],color='blue',label='cluster3')
plt.scatter(reduce_X[reduce_X['Cluster'] == 3].loc[:,'PCA1'],reduce_X[reduce_X['Cluster'] == 3].loc[:,'PCA2'],color='yellow',label='cluster4')
plt.scatter(reduce_X[reduce_X['Cluster'] == 4].loc[:,'PCA1'],reduce_X[reduce_X['Cluster'] == 4].loc[:,'PCA2'],color='orange',label='cluster5')
plt.scatter(reduce_X[reduce_X['Cluster'] == 5].loc[:,'PCA1'],reduce_X[reduce_X['Cluster'] == 5].loc[:,'PCA2'],color='purple',label='cluster6')
plt.scatter(reduce_centers[:,0],reduce_centers[:,1],color='black',marker='o',s=300)
plt.legend()
plt.show()
```

```python
# using Classification 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=100)
```

```python
from sklearn.neural_network import MLPClassifier
model1 = MLPClassifier(hidden_layer_sizes=(256,256),max_iter=500)
```

```python
model1.fit(X_train,y_train)
```

```python
print(f"accuracy score : {model1.score(X_test,y_test)}")
```
