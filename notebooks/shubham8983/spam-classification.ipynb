```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
```

```python
df = pd.read_csv("../input/sms-spam-collection-dataset/spam.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
L = LabelEncoder()
```

```python
df['v1'] = L.fit_transform(df['v1'])
```

```python
df.head()
```

```python
lem = WordNetLemmatizer()
```

```python
sentence = []
for i in range(0,len(df)):
    msg = re.sub('[^a-zA-Z]',' ',df['v2'][i])
    msg = msg.lower()
    msg = msg.split()
    msg = [lem.lemmatize(word) for word in msg if not word in set(stopwords.words('english'))]
    msg = ' '.join(msg)
    sentence.append(msg)
```

```python
sentence[0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
T = TfidfVectorizer()
```

```python
X = T.fit_transform(sentence).toarray()
y = df['v1']
```

```python
X.shape
```

```python
y.shape
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
rs = RandomOverSampler(random_state=42)
```

```python
X, y = rs.fit_resample(X,y)
```

```python
print(X.shape,y.shape)
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
MB = MultinomialNB()
```

```python
MB.fit(X_train,y_train)
```

```python
print(MB.score(X_test,y_test))
```

```python
y_pred = MB.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,f1_score
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```
