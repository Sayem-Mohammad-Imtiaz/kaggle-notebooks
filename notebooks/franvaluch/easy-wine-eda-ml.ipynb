```python
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score
from collections import Counter
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
```

```python
robust_sc = preprocessing.RobustScaler()
standard_sc = preprocessing.StandardScaler() 
minmax_sc = preprocessing.MinMaxScaler() 
```

```python
reviews = []
for i in df['quality']:
    if i >= 1 and i <= 5:
        reviews.append('1')
    elif i >= 6 and i <= 7:
        reviews.append('2')
    elif i >= 8 and i <= 10:
        reviews.append('3')
df['Reviews'] = reviews
```

```python
X = df.drop(["quality",'Reviews'],axis=1)
y = df["Reviews"]
```

```python
resultado = []
for x in [robust_sc,standard_sc,minmax_sc]:
    %time
    scaler = x.fit(X)
    X_new = x.transform(X)
    tree = DecisionTreeClassifier(random_state=42)
    tree.fit(X_new,y)
    y_pred = tree.predict(X_new)
    f1sc=f1_score(y, y_pred, average='weighted')
    rauc=(y, y_pred)
    resultado.append(f1sc)
    print("El escalado Utilizado--->",x)
    print("f1 segun el tipo de estrategia:",f1sc)
    print("----------------------------------------")
```

```python
Counter(df['Reviews'])
```

```python
over = SMOTE()
under = RandomUnderSampler()
steps = [('o', over), ('u', under)]
pipeline = Pipeline(steps=steps)
```

```python
X, y = pipeline.fit_resample(X, y)
```

```python
Counter(y)
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
```

```python
X_train=standard_sc.fit_transform(X_train)
X_test=standard_sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import cross_val_score
```

### DecisionTreeClassifier

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
dt_predict = dt.predict(X_test)
print(classification_report(y_test, dt_predict))
```

```python
scores = cross_val_score(dt, X_train, y_train, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

### RandomForestClassifier

```python
rfc = RandomForestClassifier(n_estimators=250)
rfc.fit(X_train, y_train)
pred_rfc = rfc.predict(X_test)
print(classification_report(y_test, pred_rfc))
```

```python
scores = cross_val_score(rfc, X_train, y_train, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

### SGDClassifier

```python
sgd = SGDClassifier(penalty=None)
sgd.fit(X_train, y_train)
pred_sgd = sgd.predict(X_test)
print(classification_report(y_test, pred_sgd))
```

```python
scores = cross_val_score(sgd, X_train, y_train, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

### SVC

```python
svc = SVC()
svc.fit(X_train, y_train)
pred_svc = svc.predict(X_test)
print(classification_report(y_test, pred_svc))
```

```python
scores = cross_val_score(svc, X_train, y_train, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
rbf_svc = SVC(kernel='linear')
rbf_svc.fit(X_train, y_train)
rbf_svc_res=rbf_svc.predict(X_test)
print(classification_report(y_test, rbf_svc_res))
```

```python
scores = cross_val_score(rbf_svc, X_train, y_train, cv=5)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```
