```python
data = pd.read_csv('../input/student-alcohol-consumption/student-mat.csv')
data
```

```python
object_col = [data.columns[index] for index, dtype in enumerate(data.dtypes) if dtype == 'object']
object_col
```

```python
for column in object_col:
    print(f"{column} : {data[column].unique()}")
```

```python
data['Mjob'] = data['Mjob'].apply(lambda x : 'm_' + x)
data['Fjob'] = data['Fjob'].apply(lambda x : 'f_' + x)
data['reason'] = data['reason'].apply(lambda x : 'r_' + x)
data['guardian'] = data['guardian'].apply(lambda x : 'g_' + x)
```

```python
data
```

```python
pd.get_dummies(data['Mjob'])
```

```python
dummy = pd.concat([
    pd.get_dummies(data['Mjob']),
    pd.get_dummies(data['Fjob']),
    pd.get_dummies(data['reason']),
    pd.get_dummies(data['guardian'])],
    axis = 1)

dummy
```

```python
data = pd.concat([data, dummy], axis = 1)
data.drop(['Mjob', 'Fjob', 'reason', 'guardian'], axis = 1, inplace = True)
```

```python
data
```

```python
object_col = [data.columns[index] for index, dtype in enumerate(data.dtypes) if dtype == 'object']

for column in object_col:
    print(f"{column} : {data[column].unique()}")
```

```python
encoder = LabelEncoder()

for column in object_col:
    data[column] = encoder.fit_transform(data[column])
```

```python
data.dtypes
```

```python
y = data['G3']
X = data.drop('G3', axis = 1)
```

```python
y
```

```python
X
```

```python
scaler = StandardScaler()

X = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
print(f"Model score : {model.score(X_test, y_test)}")
```

```python
y_pred = model.predict(X_test)
```

```python
print(f'Explained variance regression score : {sklearn.metrics.explained_variance_score(y_test, y_pred)}')
print(f'maximum residual error : {sklearn.metrics.max_error(y_test, y_pred)}')
print(f'Mean absolute error : {sklearn.metrics.mean_absolute_error(y_test, y_pred)}')
print(f'Mean squared error : {sklearn.metrics.mean_squared_error(y_test, y_pred)}')
print(f'r2 score : {sklearn.metrics.r2_score(y_test, y_pred)}')
```
