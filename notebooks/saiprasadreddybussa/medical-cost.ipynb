Converting categorical varibles to numerical usiing dummies

```python
data = pd.get_dummies(raw_data,columns=['sex','smoker','region'],drop_first=True)
data.head()
```

```python
data.hist('charges')
```

```python
data.charges = np.log(data.charges)
data.hist('charges');
```

```python
data.corr()
```

```python
import seaborn as sns
# correlation plot
corr = data.corr()
sns.heatmap(corr, cmap = 'Wistia', annot= True);
```

```python
X = data.copy()
```

```python
del X['charges']
X.shape
```

```python
#X = X.drop(['region_northwest','region_southeast','region_southwest'],axis=1)
#X.shape
```

```python
y=data['charges']
y.shape
```

Splitting the data to train and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
pred = model.predict(X_test)
```

```python
sse=np.sum((y_test-pred)**2)
sse
```

```python
sst = np.sum((pred-np.mean(y_train))**2)
sst
```

```python
r2=1-(sse/sst)
r2
```
