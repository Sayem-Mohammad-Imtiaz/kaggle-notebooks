# Hazırlık
Öncelikle gerekli kütüphaneleri ekleyelim.

```python

import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Kurulum Tamamlandı.")

```

Daha sonra veri setimizi csv dosyasından okuyalım. Bunu yaparken pandas kütüphanesinden yardım alarak csv formatındaki verileri dataFrame formatına dönüştürüyoruz. DataFrame, bir SQL tablosu gibi düşünülebilir içerdiği satır ve sütunlar ile verilerimizi düzenli bir şekilde tutmaya yarar. 

```python
weatherww2_filepath = "../input/weatherww2/Summary of Weather.csv"
weatherww2_data = pd.read_csv(weatherww2_filepath,  index_col="Date", parse_dates=True)
```

Görmüş olduğunuz gibi pd.read_csv fonksiyonu 3 parametre aldı bunlardan birincisi okumak istediğimiz dosyanın yolu, ikincisi indeksleme yapılacak sütun, üçüncüsü ise tarih ile alakalı alanların tip dönüşümünü sağlar.

## Veriyi İnceleyelim

Bu işlemleri gerçekleştirdikten sonra describe() fonksiyonunu kullanarak verilerimizin özetini görelim.

```python
weatherww2_data.describe()
```

describe() fonksiyonu sonucunda oluşan tabloda veri setimiz ile alakalı genel bilgileri inceleyebilirsiniz örneğin MinTemp için en küçük değer -38.33 olduğu gözüküyor. Bunun gibi veri setinize genel bir bakış için oldukça güzel bir fonksiyon. Ayrıca verilerin bir kısmını görmek için tail ve head fonksiyonları kullanılabilir. head() fonksiyonuna parametre verilmezse veri setindeki ilk 5 satırı gösterecektir ya da görmek istediğimiz satır sayısını parametre olarak verip istediğiniz kadar satırı inceleyebilirsiniz.

```python
weatherww2_data.head(10)
```

Veri setimizdeki ilk 10 satıra bakacak olursak NaN şeklinde değerler görmekteyiz bunun anlamı Not a Number yani sayısal olmayan değer. Bu notebookta çizgi grafiği kullanacağımız için sayısal olmayan değerleri kullanmayacağız.

## Seaborn Çizgi Grafiği 

Seaborn kütüphanesindeki lineplot() fonksiyonunu kullanarak veri setimizdeki birkaç verinin grafiğini çizdirelim. Bunun için alt blokta da görebileceğiniz üzere figür boyutunu, figürün başlığını ve veri setimizdeki hangi sütunun grafiğini çizdirmek istediğimizi seçtik.

```python

plt.figure(figsize=(12,6))
sns.lineplot(data=weatherww2_data["MeanTemp"])
plt.title("1940-1945 Yılları Arasındaki Günlük Ortalama Sıcaklıklar")
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=weatherww2_data["MaxTemp"])
plt.title("1940-1945 Yılları Arasındaki Günlük Maksimum Sıcaklıklar")
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=weatherww2_data["MinTemp"])
plt.title("1940-1945 Yılları Arasındaki Günlük Maksimum Sıcaklıklar")
```

Günlük ortalama sıcaklık, günlük maksimum ve günlük minimum sıcaklık verilerinin çizgi grafiklerini çizdirdik. Siz de Seaborn kullanarak grafikler çizdirebilir ve verilerinizi yorumlayabilirsiniz. :)
