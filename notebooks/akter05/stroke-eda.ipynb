# 2. Distribution Plot

# 2.1 Average Glucose Level

```python
fig = plt.figure(figsize=(8,8))
plt.title('Distribution of Average Glucose Level', fontsize = 25)
sns.distplot(df1.avg_glucose_level, label="avg_glucose_level", kde= False)
plt.legend()
```

# 2.2 No Stroke vs Stroke by Avg. Glucose Level

```python
plt.figure(figsize=(12,10))

sns.distplot(df1[df1['stroke'] == 0]["avg_glucose_level"], color='green') # No Stroke - green
sns.distplot(df1[df1['stroke'] == 1]["avg_glucose_level"], color='red') # Stroke - Red

plt.title('No Stroke vs Stroke by Avg. Glucose Level', fontsize=15)
plt.xlim([30,330])
plt.show()
```

# 2.3 BMI

```python
fig = plt.figure(figsize=(8,8))
plt.title('Distribution of BMI', fontsize = 25)
sns.distplot(df1.avg_glucose_level, label="bmi", kde= False)
plt.legend()
```

# 2.4 No Stroke vs Stroke by BMI

```python
plt.figure(figsize=(12,10))

sns.distplot(df1[df1['stroke'] == 0]["bmi"], color='green') # No Stroke - green
sns.distplot(df1[df1['stroke'] == 1]["bmi"], color='red') # Stroke - Red

plt.title('No Stroke vs Stroke by BMI', fontsize=15)
plt.xlim([10,100])
plt.show()
```

# 2.5 Pair Plot

```python
fig = plt.figure(figsize=(10,10))
sns.pairplot(df1, hue = 'stroke')
plt.show()
```
