```python
import pandas as pd
import seaborn as sns
df =  pd.read_csv('../input/Pokemon.csv', index_col='#')

pkmn = df.groupby(['Legendary', 'Generation']).mean()[['Attack', 'Defense']]
pkmn.plot.bar(stacked=True)

```
