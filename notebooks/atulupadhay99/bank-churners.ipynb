```python
pd.set_option('max_columns',None)
df = pd.read_csv('/kaggle/input/credit-card-customers/BankChurners.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop(['CLIENTNUM','Card_Category'],axis=1)
```

```python
df2=df.copy()
df2.head()
```

```python
df.Marital_Status.unique()
```

```python
categorical = df2.select_dtypes(include = 'object').columns
categorical
```

```python
dummies = pd.get_dummies(df2.loc[:,['Attrition_Flag', 'Gender', 'Education_Level', 'Marital_Status',
       'Income_Category']])
dummies.head()
```

```python
dummies.columns
```

```python
dummies.drop(['Attrition_Flag_Existing Customer',
       'Gender_F','Education_Level_Uneducated',
       'Marital_Status_Single','Income_Category_Less than $40K'],axis=1,inplace = True)
```

```python
dummies.head()
```

```python
df3 = pd.concat([df2,dummies],axis=1)
```

```python
df3.head()
```

```python
df3.drop(categorical,axis=1,inplace = True)
```

```python
df3.head()
```

```python
df3.info()
```

```python
df4 = df3.copy()
```

```python
X = df3.drop('Attrition_Flag_Attrited Customer',axis=1)
y = df3['Attrition_Flag_Attrited Customer']


from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

for cols in X.columns:
    df3[cols] = scaler.fit_transform(df3[[cols]])
df3.head()
```

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf',C=1,gamma='scale')

model.fit(X_train,y_train).predict(X_test)
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cvs = cross_val_score(model,X,y,cv = 10)
```

```python
cvs
```

```python
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()

model1.fit(X_train,y_train).predict(X_test)
model1.score(X_test,y_test)
```

```python
cvs1 = cross_val_score(model1,X,y,cv = 10)
```

```python
cvs1
```

```python
X = df4.drop('Attrition_Flag_Attrited Customer',axis=1)
y = df4['Attrition_Flag_Attrited Customer']


from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3)

from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier()

model2.fit(X_train,y_train).predict(X_test)
model2.score(X_test,y_test)
```

```python
cvs2 = cross_val_score(model2,X,y,cv=10)
cvs2
```
