```python
df.isnull().sum()
```

```python
df['Arrival Delay in Minutes'].fillna((df['Arrival Delay in Minutes'].mean()),axis=0,inplace = True)
```

```python
df.isnull().sum()
```

```python
df2 = df.copy()
df2.head()
```

```python
df2.drop(['Unnamed: 0', 'id','Gate location'],axis=1,inplace = True)
```

```python
categorical = df2.select_dtypes(include = 'object').columns
categorical
```

```python
dummies = pd.get_dummies(df.loc[:,['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']])
dummies.head()
```

```python
dummies.drop(['Gender_Female','Customer Type_Loyal Customer','Type of Travel_Business travel',
             'Class_Business','satisfaction_neutral or dissatisfied'],axis=1,inplace = True)
```

```python
dummies
```

```python
df3 = pd.concat([dummies,df2],axis = 1)
df3.drop(categorical,axis = 1,inplace = True)
df3.head()
```

```python
columns = ['Gender_Male', 'Customer Type_disloyal Customer',
       'Type of Travel_Personal Travel', 'Class_Eco', 'Class_Eco Plus',
       'Age', 'Flight Distance',
       'Inflight wifi service', 'Departure/Arrival time convenient',
       'Ease of Online booking', 'Food and drink', 'Online boarding',
       'Seat comfort', 'Inflight entertainment', 'On-board service',
       'Leg room service', 'Baggage handling', 'Checkin service',
       'Inflight service', 'Cleanliness', 'Departure Delay in Minutes',
       'Arrival Delay in Minutes']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

for cols in columns:
    df3[cols] = scaler.fit_transform(df3[[cols]])
df3.head()
```

```python
X = df3.drop('satisfaction_satisfied',axis = 1)
y = df3.satisfaction_satisfied
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```
