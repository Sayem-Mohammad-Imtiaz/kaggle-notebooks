# Create Model
We will try multiple possible models and see which gives best possible accuracy on testing data.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
state = 0
```

```python
tester = {
    "Logistic Regression": {
        "model": LogisticRegression(random_state = state, multi_class = "auto"),
        "params": {
            "C": [0.1, 0.01, 0.001, 0.0001, 1, 10],
            "solver": ["liblinear", "sag", "lbfgs", "newton-cg", "sag", "saga"]
        }
    },
    
    "K Nearest Neighbors": {
        "model": KNeighborsClassifier(),
        "params":{
            "n_neighbors": list(range(5, 15))
        }
    },
    
    "Random Forest Classifier": {
        "model": RandomForestClassifier(random_state = state),
        "params": {
            "n_estimators": [1, 5, 10, 50, 100, 200, 150, 125],
            "criterion": ["gini", "entropy"]
        }
    },
    
    "SVM": {
        "model": SVC(random_state = state),
        "params": {
            "degree": ["linear", "poly", "rbf", "sigmoid"],
            "C": [0.1, 0.01, 0.001, 0.0001, 1, 10],
            "degree": list(range(3, 10))
        }
    },
    
    "Naive Bayes":{
        "model": GaussianNB(),
        "params": dict()
    }
}
```

Logistic Regression training causes warning which they must be ignored.

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
scores = list()

for name, model in tester.items():
    clf = GridSearchCV(model["model"], model["params"], cv = 10, return_train_score = False)
    clf.fit(x_train, y_train)
    test = {
        "Model": name,
        "Best Score": clf.best_score_,
        "Best Prams": clf.best_params_
    }
    
    clf = clf.best_estimator_
    
    test["Test Score"] = accuracy_score(clf.predict(x_test), y_test)
    scores.append(test)
    print(f"{name} trained")
```

```python
# No scaling
for i in scores:
    print(i)
```

Observations:
1. Naive Bayes performs worst at 56%.
2. K Nearest Neighbors is next with 81.2%.
3. Logistic Regression is next with 81.6%.
4. Random Forest is next with 83.8%.
5. SVM is next with 84.2%.<br>

Top two models are Random Forest and SVM, we will work upon them.

# Try The Same With Scaling
While currently all features are min-max scaled, we will try again using standard scaling too see if there is any change in accuracy.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
scores = list()

for name, model in tester.items():
    clf = GridSearchCV(model["model"], model["params"], cv = 10, return_train_score = False)
    clf.fit(x_train_scaled, y_train)
    test = {
        "Model": name,
        "Best Score": clf.best_score_,
        "Best Prams": clf.best_params_
    }
    
    clf = clf.best_estimator_
    
    test["Test Score"] = accuracy_score(clf.predict(x_test_scaled), y_test)
    scores.append(test)
    print(f"{name} trained")
```

```python
# with scaling
for i in scores:
    print(i)
```

Here accuracies become:
1. Naive Bayes 55.4%.
2. K Nearest Neighbors 80.4%.
3. Logistic Regression 81.4%.
4. SVM with 83.6%.
5. Random Forest with 83.8%.<br> 

We tried to optimise SVM using GridSearchCV and the lowest possible settings gave best accuracy and scaling further decrease it, we will try to see if RandomForest can be optimised using PCA.

# Improving The Model
We will try using PCA to improve accuracy.

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 10)
x_pca = pca.fit_transform(x_train_scaled)
pca.explained_variance_ratio_
```

```python
clf = RandomForestClassifier()
clf.fit(x_pca, y_train)

x_pca2 = pca.transform(x_test_scaled)
y_pred = clf.predict(x_pca2)
accuracy_score(y_pred, y_test)
```

```python
scores = list()

for i in range(1, 32):
    pca = PCA(n_components = i)
    x_pca = pca.fit_transform(x_train_scaled)
    
    clf = GridSearchCV(RandomForestClassifier(random_state = state), {
        "n_estimators": [1, 5, 10, 50, 100, 200, 150, 125],
        "criterion": ["gini", "entropy"]
    }, cv = 10, return_train_score = False)
    clf.fit(x_pca, y_train)
    
    test = dict()
    test["model"] = clf.best_estimator_
    clf = clf.best_estimator_

    x_pca2 = pca.transform(x_test_scaled)
    y_pred = clf.predict(x_pca2)
    test["accuracy score"] = accuracy_score(y_pred, y_test)
    scores.append(test)
    
    print(f"Iteration {i} over")
```

```python
for i in scores:
    print(i)
```

```python
accuracies = list()

for val in scores:
    accuracies.append(val["accuracy score"])

plt.plot(range(len(scores)), accuracies)
plt.show()
```

Dictionary contains models along with their accuracy_scores. Highest scoring model will be the final model.

# Final Model

```python
max_accuracy = max(accuracies)
```

```python
max_accuracy
```

```python
for i in scores:
    if max_accuracy == i["accuracy score"]:
        best = i
```

```python
best
```

This is the most accurate model.
