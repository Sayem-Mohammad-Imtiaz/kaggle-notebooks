```python
df.drop(['CLIENTNUM','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'], axis=1)
```

```python
existing,attrited = df['Attrition_Flag'].value_counts()
plt.pie([existing,attrited], explode=(0.1,0),labels=['Existing Customer', 'Attrited Customer'],colors=['#009ACD', '#ADD8E6'],autopct='%1.2f%%')
```

```python
bplot = plt.boxplot([df.loc[df['Attrition_Flag'] == 'Existing Customer']['Customer_Age'],df.loc[df['Attrition_Flag'] == 'Attrited Customer']['Customer_Age']],'red','blue',patch_artist=True)
for patch, color in zip(bplot['boxes'], ['lightblue','lightgreen']):
    patch.set_facecolor(color)
plt.legend([bplot["boxes"][0], bplot["boxes"][1]],['Existing Customer','Attrited Customer'])
plt.show()
```

```python
exist_female,exist_male = df[df['Attrition_Flag'] == 'Existing Customer']['Gender'].value_counts()
attrited_female,attrited_male = df[df['Attrition_Flag'] == 'Attrited Customer']['Gender'].value_counts()
plt.bar(['Existing Male','Existing Female','Attrited Male','Attrited Female'],[exist_male,exist_female,attrited_male,attrited_female],color=['#009ACD','#d87cd8','lightblue','pink'])
plt.show()
```

```python
exist_data = df[df['Attrition_Flag'] == 'Existing Customer']['Education_Level'].value_counts()
attrited_data = df[df['Attrition_Flag'] == 'Attrited Customer']['Education_Level'].value_counts()
exist_label = list(map(lambda x: 'Existing '+ x,exist_data.index.tolist()))
attrited_label = list(map(lambda x: 'Attrited '+ x,attrited_data.index.tolist()))
data = list(exist_data) + list(attrited_data)
plt.bar(exist_label+attrited_label,data,color=['#009ACD','#d87cd8','lightblue','pink'],align='edge')
plt.xticks(exist_label+attrited_label, exist_label+attrited_label, rotation='vertical')
plt.show()
```

```python
plt.pie(data,labels=exist_label+attrited_label,autopct='%1.0f%%')
plt.show()
```

```python
exist_data = df[df['Attrition_Flag'] == 'Existing Customer']['Income_Category'].value_counts()
attrited_data = df[df['Attrition_Flag'] == 'Attrited Customer']['Income_Category'].value_counts()
exist_label = list(map(lambda x: '(Existing) '+ x,exist_data.index.tolist()))
attrited_label = list(map(lambda x: '(Attrited) '+ x,attrited_data.index.tolist()))
data = list(exist_data) + list(attrited_data)
plt.bar(exist_label+attrited_label,data,color=['#009ACD','#d87cd8','lightblue','pink'],align='edge')
plt.xticks(exist_label+attrited_label, exist_label+attrited_label, rotation='vertical')
plt.show()
```
