## Scatter Plot

With the help of FacetGrid, which is imported from Seaborn. <br>  <br>Using "hue='Patient_Survival_morethan_5Years'", we can differentiate the Classification between 'Patient_age' and 'Auxilary_Node'. <br> <br> Most of the Points are overlapping it is some what difficult to come for confusion so we plot for the other 3 attirbutes using Pair Plots

## 1D Scatter_Plot

Below we plot the 1D Scatter Plot to find any conclusion even in this most of the Data Points are overlapping

```python
hbs_long_survived = hbs.loc[hbs['Patient_Survival_morethan_5Years']==1]
hbs_short_survived = hbs.loc[hbs['Patient_Survival_morethan_5Years']==2]
plt.plot(hbs_long_survived['Auxilary_Node'], np.zeros_like(hbs_long_survived['Auxilary_Node']) , 'o')
plt.plot(hbs_short_survived['Auxilary_Node'], np.zeros_like(hbs_short_survived['Auxilary_Node']) , 'o')
plt.xlabel('Auxilary_Node')
plt.title('Scatter Plot of Survival vs Auxilary_Node')
```

## PDF of Auxilary_Node

It is the PDF of "Auxilary_Nodes"In this Plot we can see that, Auxilary_Nodes Less than 0 means Survival_morethan_5Years are high, <br> <br>
Auxilary_Nodes are Greater than 0 and Less than 3 then Survival_morethan_5Years are Moderate, <br> <br>
Auxilary_Nodes are Greater than 3 and Less than 5 then Survival_morethan_5Years are less

```python
sns.FacetGrid(hbs, hue='Patient_Survival_morethan_5Years', size=6).map(sns.distplot, 'Auxilary_Node').add_legend()
plt.title('Patient_Survival_morethan_5Years vs Auxilary_Node')
```

## CDF of Long_Survived

From Below "CDF for Long_Survives" we can come to conclusion that if the Auxilary nodes are less means survival chances are more. <br> <br> In this case Auxillary nodes < 5 there are 82% of Survival_morethan_5Years. We can observe that if the Auxillary nodes are increasing Survival rate is decreasing

```python
counts, bin_edges = np.histogram(hbs_long_survived['Auxilary_Node'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(['1','2'])
plt.xlabel('Auxilary_Node')
plt.title('CDF of Long_Survived against Auxilary_Nodes')
```

## CDF of Short_survived

In the below "CDF of survived_lessthan_5Years", we can observe there is a 100% survival even Auxillary_Nodes > 40

```python
counts, bin_edges = np.histogram(hbs_short_survived['Auxilary_Node'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
plt.legend(['1','2'])
plt.xlabel('Auxilary_Node')
plt.title('CDF of Short_Survived against Auxilary_Nodes')
```
