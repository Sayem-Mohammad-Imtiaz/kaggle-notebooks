```python
dataFrame = pd.read_csv('/kaggle/input/customer-segmentation-tutorial-in-python/Mall_Customers.csv')
dataFrame.head(10)
```

```python
## let map Male => 0
## and Female => 1

mp1 = {
    'Male' : 0,
    'Female' : 1
}

dataFrame['Gender'] = dataFrame['Gender'].map(mp1)
```

```python
## lets convert the values np.array
ConsumerVector = np.array(dataFrame[['Gender', 'Age', 'Annual Income (k$)','Spending Score (1-100)']])
```

```python
## since we dont have any idea about no. of clusters we will use 
## Affinity propagation
import sklearn
from sklearn.cluster import AffinityPropagation
clustering = AffinityPropagation(random_state=5).fit(ConsumerVector)
```

```python
#### let see the cluster centres so formed
clusterCentres = clustering.cluster_centers_
i = 0
for mem in clusterCentres:
    i += 1
    gender = mem[0]
    age = mem[1]
    annualIncome = mem[2]
    spendingScore = mem[3]
    
    print ('======================================================================================')
    print ('cluster Centre : '+ str(i) + ' has the following specs :' )
    print ('======================================================================================')
    print ('Age ==>  {}'.format(age))
    print  ('Gender ==> {}'.format(gender))
    print('Spending Score ==>  {}'.format(spendingScore))
    print('Annual income(k) ==> {}'.format(annualIncome))
    print ('======================================================================================')
```

```python
##############################################################################################
# Notes:                                                                                     #
##############################################################################################
# We have 12 cluster centres described as follows:                                           #
# 1.) cluster 1 females around 54 with 28 k income spending score 14                         #
# 2.) cluster 2 males around 29 with income 28 k  spending score 82  *                       #
# 3.) cluster 3. females around  40 income 29 k spending score 31                            #
# 4.) cluster 4. females around 27 income 46 k spending score 46                             #
# 5.) cluster 5. males around 63 48 k income spending score 51                               #
# 6.) cluster 6. males around 48 54 k income spending score 54                               #
# 7.) cluster 7. females around 23 62 k income spending score 41 *                           #
# 8.) cluster 8 female around 32 77 k income spending score   74 **                          #
# 9.) cluster 9 male around 34 income 78 k spending score  90**                              #
# 10.) cluster 10 males around income 78 k  spending score 17*****                           #
# 11.) cluster 11females around 41 income 103 spending score 17******                        #
# 12.) cluster 12 females around 36 income 103 k spending score 85                           #
##############################################################################################

##############################################################################################
# Remarks:                                                                                   #
##############################################################################################
# for both males and females spike in expediture is noticed around 29 - 36                   #
# Consumers in their early 40's is an untapped segment they have the money but are'nt speding#
# much                                                                                       #
##############################################################################################
```

```python
## let add cluster colum to our data frame
dataFrame['Cluster number'] = clustering.predict(ConsumerVector)
```

```python
## let sort our dataframe by cluster number
dataFrame.sort_values(by='Cluster number', inplace= True)
```
