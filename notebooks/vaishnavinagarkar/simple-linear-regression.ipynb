```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'../input/indian-food-101/indian_food.csv')
```

```python
df.head() 
```

```python
x=df['prep_time'] 
x
```

```python
y=df['cook_time']
y
```

```python
plt.scatter(x,y)
plt.show() 
```

```python
x_mean=np.mean(x)
x_mean
```

```python
y_mean=np.mean(y)
y_mean
```

```python
m=sum((x-x_mean)*(y-y_mean))/sum((x-x_mean)*(x-x_mean))
```

```python
m
```

```python
c=y_mean-m*x_mean
c
```

```python
def predict_time(x):
    y_predicted=0.09586901450902668*x+31.54732147809569
    return y_predicted 
y_predicted =predict_time(x)
y_predicted
```

```python
plt.scatter(x,y) 
plt.plot(x,y_predicted, color='black')  
plt.show()  
```

```python
r_square=1-sum((y-y_predicted)*(y-y_predicted))/sum((y-y_mean)*(y-y_mean))
r_square
```
