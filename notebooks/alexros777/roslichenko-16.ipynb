```python
df = pd.read_csv('/kaggle/input/videogamesales/vgsales.csv')
df
```

```python
import statistics as stat

mediana_NA_Sales=np.median(df['NA_Sales'])
print("Медиана NA_Sales: ", mediana_NA_Sales)

mediana_EU_Sales=np.median(df['EU_Sales'])
print("Медиана EU_Sales: ", mediana_EU_Sales)

mediana_JP_Sales=np.median(df['JP_Sales'])
print("Медиана JP_Sales: ", mediana_JP_Sales)

mode_NA_Sales=stat.mode(df['NA_Sales'])
print("Мода NA_Sales: ", mode_NA_Sales)

mode_EU_Sales=stat.mode(df['EU_Sales'])
print("Мода EU_Sales: ", mode_EU_Sales)

mode_JP_Sales=stat.mode(df['JP_Sales'])
print("Мода JP_Sales: ", mode_JP_Sales)

mean_NA_Sales=np.mean(df['NA_Sales'])
print("Средняя NA_Sales: ", mean_NA_Sales)

mean_EU_Sales=np.mean(df['EU_Sales'])
print("Средняя EU_Sales: ", mean_EU_Sales)

mean_JP_Sales=np.mean(df['JP_Sales'])
print("Средняя JP_Sales: ", mean_JP_Sales)

max_NA_Sales=np.max(df['NA_Sales'])
print("Максимум NA_Sales: ", max_NA_Sales)

max_EU_Sales=np.max(df['EU_Sales'])
print("Максимум EU_Sales: ", max_EU_Sales)

max_JP_Sales=np.max(df['JP_Sales'])
print("Максимум JP_Sales: ", max_JP_Sales)

min_NA_Sales=np.min(df['NA_Sales'])
print("Минимум NA_Sales: ", min_NA_Sales)

min_EU_Sales=np.min(df['EU_Sales'])
print("Минимум EU_Sales: ", min_EU_Sales)

min_JP_Sales=np.min(df['JP_Sales'])
print("Минимум JP_Sales: ", min_JP_Sales)
```
