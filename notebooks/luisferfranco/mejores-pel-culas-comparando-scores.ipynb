```python
df = pd.read_csv('/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv')
df.IMDB_Rating *= 10
df.head().T
```

```python
df['favoritas'] = df.apply(lambda x: math.sqrt((x.Meta_score - 100)**2 + 
                                               (x.IMDB_Rating - 100)**2), 
                           axis=1)
df['esnobismo'] = df.apply(lambda x: math.sqrt((x.Meta_score - 100 )**2 + 
                                               (x.IMDB_Rating)**2), 
                           axis=1)
df['churros'  ] = df.apply(lambda x: math.sqrt((x.Meta_score)**2 + 
                                               (x.IMDB_Rating)**2), 
                           axis=1)
df['palomeras'] = df.apply(lambda x: math.sqrt((x.Meta_score)**2 + 
                                               (x.IMDB_Rating - 100)**2), 
                           axis=1)
```

```python
import matplotlib.pyplot as plt
```

```python
df.sort_values(by='favoritas', inplace=True)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(20,20))

plt.xlim(20, 105)
plt.ylim(20, 105)
#plt.fill([50, 100, 50, 50], [50, 100, 100, 50], c='#f4f1de50')
#plt.fill([50, 100, 100, 50], [50, 50, 100, 50], c='#74c69d50')

plt.axhline(60, c='#cccccc', ls='--', lw=3)
plt.axvline(60, c='#cccccc', ls='--', lw=3)

plt.scatter(x=df.Meta_score, y=df.IMDB_Rating, c='steelblue', marker='o')
plt.text(20, 100, 'PALOMERAS', verticalalignment='top')
plt.text(100, 100, 'LAS MEJORES', verticalalignment='top', horizontalalignment='right')
plt.text(100, 20, 'SUPER SNOB', horizontalalignment='right')
plt.text(20, 20, 'MALAS')


plt.plot((20, 100), (20, 100), 'r:', lw=2)
plt.title('Comparación de Calificaciones Críricos vs. Usuarios de IMDB')
plt.xlabel('CRITICOS (META)')
plt.ylabel('USUARIOS DE IMDB')

peli = df[['Series_Title', 
           'favoritas', 
           'Meta_score', 
           'IMDB_Rating']].sort_values(by='favoritas').iloc[0]
plt.scatter(x=peli.Meta_score, y=peli.IMDB_Rating,
            c='gold', marker='*', 
            edgecolor='k', s=300)
plt.text(peli.Meta_score, peli.IMDB_Rating-2,
         peli.Series_Title, horizontalalignment='center')

peli = df[['Series_Title', 
           'palomeras', 
           'Meta_score', 
           'IMDB_Rating']].sort_values(by='palomeras').iloc[0]
plt.scatter(x=peli.Meta_score, y=peli.IMDB_Rating,
            c='green', marker='d', 
            edgecolor='k', s=300)
plt.text(peli.Meta_score, peli.IMDB_Rating-2,
         peli.Series_Title, horizontalalignment='center')

peli = df[['Series_Title', 
           'churros', 
           'Meta_score', 
           'IMDB_Rating']].sort_values(by='churros').iloc[0]
plt.scatter(x=peli.Meta_score, y=peli.IMDB_Rating,
            c='r', marker='X', 
            edgecolor='k', s=300)
plt.text(peli.Meta_score, peli.IMDB_Rating-2,
         peli.Series_Title, horizontalalignment='center')

peli = df[['Series_Title', 
           'esnobismo', 
           'Meta_score', 
           'IMDB_Rating']].sort_values(by='esnobismo').iloc[0]
plt.scatter(x=peli.Meta_score, y=peli.IMDB_Rating,
            c='purple', marker='P', 
            edgecolor='k', s=300)
plt.text(peli.Meta_score, peli.IMDB_Rating-2,
         peli.Series_Title, horizontalalignment='center')
```
