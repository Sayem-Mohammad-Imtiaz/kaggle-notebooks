# To Change Column Names

```python
data.rename(columns={'Order Date & Time': 'order_date'}, inplace=True)
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(' ', '_')
data.columns = data.columns.str.replace('(', '' )
data.columns = data.columns.str.replace(')', '')
data.head()
```

```python
data['order_date'] = pd.to_datetime(data.order_date)
data['year'] = data.order_date.dt.year
data['month'] = data.order_date.dt.month_name()
data['date'] = data.order_date.dt.day
data['day'] = data.order_date.dt.day_name()
```

```python
data.describe(include='all')
```

```python
data.isnull().sum().nlargest() # to check null values
```

# Order Status - Findings in Data Frame
* There are only 3 Unique Value so we start with this one to explore our data

```python
order = data.order_status.value_counts()
total_order = data.order_status.count()
print('Total number of ** ORDER RECEIVED ** during the period: ', total_order)
print('Total number of ** COMPLETED ORDER ** during the period: ', order.values[0])
print('Total number of ** RETURNED ORDER ** during the period: ', order.values[1])
print('Total number of ** CANCELLED ORDER ** during the period: ', order.values[2])
```

```python
plt.figure(figsize=(20,7))

plt.subplot(1, 2, 1)
# sns.countplot(x='order_status', data=data,)
plt.bar(order.index, order.values)
plt.title("ORDER ANALYSIS", fontsize=40, color='blue')
plt.xlabel('Order Status')
plt.ylabel('Order QTY')


plt.subplot(1, 2, 2)
plt.pie(order.values, labels=order.index)

plt.show()

ax = sns.catplot(x='year', col='order_status',data=data, kind='count',linewidth=1.5);
```

**Facts about Order Status**
* there are **19239 total orders** out of which **17948 is Completed** i.e., 93.29%
* and **1142 orders are returned** i.e, 5.94%
* further **149 orders are cancelled** i.e, 0.77%

**Observation about Order Analysis**
**its mean that in **every 100 orders** there will be only **6 Retunred** and **1 order will be cancelled**

## Top Selling Months & Days

```python
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
day_order = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
data['month'] = data.month.str[0:3]
data['day'] = data.day.str[0:3]
```

```python
plt.figure(figsize=(20, 8))
plt.title('Month Wise Number of Orders', fontsize=30, color='purple')
splot = sns.countplot(x="month", data=data, order=month_order)

for p in splot.patches:
        splot.annotate(format(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')
```

```python
plt.figure(figsize=(15, 6))
plt.title('Day Wise Number of Orders', fontsize=30, color='Green')
splot = sns.countplot(x="day", data=data, order=day_order)

for p in splot.patches:
        splot.annotate(format(p.get_height()), (p.get_x() + p.get_width() / 2, p.get_height()), ha = 'center', va = 'center', xytext = (0, 10), textcoords = 'offset points')

```

```python
data['mm'] = data.order_date.dt.month
```

```python
df = data[['year', 'mm']].value_counts().reset_index()
df.columns = ['year', 'mm', 'title']
df = df[['mm', 'year', 'title']].sort_values('mm')
df = df.pivot('mm', 'year', 'title')
df = df.fillna(0)

f, ax = plt.subplots(figsize=(15,10))
sns.heatmap(df, annot=True, linewidths=1, ax=ax, fmt='n', yticklabels=month_order)
plt.show()
```
