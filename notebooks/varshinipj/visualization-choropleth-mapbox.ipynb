THREE IMPORTANT VISUALIZATIONS USING CHOROPLETH MAPBOX

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import plotly.express as px
import warnings
warnings.simplefilter("ignore")

import json
```

```python
data = pd.read_csv("../input/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv")
data.head()
```

```python
data.shape
```

```python
data = data.drop(32)
```

I removed the UT 'Ladakh' because my json file doesnot have the features of the UT.

```python
india_states = json.load(open("../input/geojson-for-indian-states/states_india.geojson", "r"))
```

```python
state_id_map = {}
for feature in india_states["features"]:
    feature["id"] = feature["properties"]["state_code"]
    state_id_map[feature["properties"]["st_nm"]] = feature["id"]
```

Mapping the state name and the id

```python
data["id"] = data["State/UTs"].apply(lambda x: state_id_map[x])
```

# ACTIVE CASES IN INDIA

```python
fig = px.choropleth_mapbox(
    data,
    locations="id",
    geojson=india_states,
    color="Active",
    hover_name="State/UTs",
    hover_data=["Active"],
    title="ACTIVE COVID-19 CASES IN INDIA",
    mapbox_style="stamen-watercolor",
    center={"lat": 24, "lon": 78},
    zoom=3,
    opacity=0.5
)
fig.show()
```

# DEATHS IN INDIA

```python
fig = px.choropleth_mapbox(
    data,
    locations="id",
    geojson=india_states,
    color="Deaths",
    hover_name="State/UTs",
    hover_data=["Deaths"],
    title="DEATHS DUE TO COVID-19 IN INDIA",
    mapbox_style="stamen-terrain",
    center={"lat": 24, "lon": 78},
    zoom=3,
    opacity=0.5,
)
fig.show()
```

# RECOVERIES IN INDIA:

```python
fig = px.choropleth_mapbox(
    data,
    locations="id",
    geojson=india_states,
    color="Discharged",
    hover_name="State/UTs",
    hover_data=["Discharged"],
    title="COVID-19 RECOVERIES IN INDIA",
    mapbox_style="open-street-map",
    center={"lat": 24, "lon": 78},
    zoom=3,
    opacity=0.5,
)
fig.show()
```
