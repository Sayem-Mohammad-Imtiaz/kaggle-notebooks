```python
import numpy as np
import pandas as pd
from scipy import stats

from sklearn.metrics import mean_squared_error
from math import sqrt

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../input/insurance/insurance.csv')
data
```

```python
# sns.scatterplot(x='bmi',y='charges',data=data,palette='magma',hue='smoker')
sns.lmplot(x="bmi", y="charges", hue="smoker", data=data, palette = 'magma', height = 8)
plt.show()
```

```python
# Smoker correlation
smoker = data[data.smoker == 'yes']
r,_ = stats.pearsonr(
    x=smoker.bmi,
    y=smoker.charges
)

r
```

```python
len(smoker)
```

```python
# Non-Smoker correlation
nonsmoker = data[data.smoker == 'no']
r,_ = stats.pearsonr(
    x=nonsmoker.bmi,
    y=nonsmoker.charges
)

r
```

RMSE of smokers

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(
    x=stats.zscore(smoker.bmi),
    y=stats.zscore(smoker.charges)
)

def expected(x):
    return intercept + slope*x

expected_ys = list(map(
    expected,
    list(stats.zscore(smoker.charges))
))

sqrt(mean_squared_error(stats.zscore(smoker.charges), expected_ys))
```

```python
sns.lmplot(data=smoker, y='charges', x='bmi')
plt.show()
```

Residual plot (smokers)

```python
sns.residplot(
    x=smoker.bmi,
    y=stats.zscore(smoker.charges),
    lowess=True
)
plt.show()
```
