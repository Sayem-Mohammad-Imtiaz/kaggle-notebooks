```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#To import the dataset by using pandas
df=pd.read_csv("/kaggle/input/students-performance-in-exams/StudentsPerformance.csv")
```

```python
""" to find the relation ship between different numerical variable is done by pair plot
the scatterplots are classified on the basis of gender,which is shown as different colours
from this we can see that the math score,reading score,writing score etc takes a bell curve ie probability density fuction"""

sns.pairplot(df,hue="gender")
```

```python
#to get the description about data
df.describe()
```

```python
#to check if there is null values
df.isnull().sum()
```

```python
#to get the list of all the unique values present in each categorical variables
[df[i].unique() for i in df.columns if i not in ["math score","reading score","writing score"]]
```

```python
# to get the graphical relation between the gender and reading score
""" from this plot we can conclude some simple facts that
>female students have the least score for reading
>most of the male students reading score is between 45 and 85 
>most of the female students score is between 55 and 90
>more female students has scored more than 90 compared to male"""
sns.histplot(df,x="gender",y="reading score")
```

```python

"""we can conclude that
>group E and A have a least score comparitively higher than the other groups"""

sns.histplot(df,x="race/ethnicity",y="math score")
```
