That's it. So, we rename it. I don't like Dt_Customer either.

```python
df=df.rename(columns={" Income ":"Income",
                   "Dt_Customer":"Date_of_Enrollment"})
df.head()
```

```python
df['Income'].head()
```

We see that 24 observations of Income feature are null.

Now we have to decide how are we going to deal with them. 

I choose to replace the missing values of Income feature using the mean function.

```python
df['Income'] = df['Income'].str.replace('$', '')
df['Income'] = df['Income'].str.replace(',', '')
df['Income'].head()
```

```python
avg_income=df['Income'].astype('float').mean()
print("The mean of Income column is:",avg_income)
```

Replace the empty values with the mean of column.

```python
df["Income"].replace(np.nan,avg_income,inplace=True)
```

Then we check out.

```python
missing_data2=df['Income'].isnull()
missing_data2.value_counts()
```
