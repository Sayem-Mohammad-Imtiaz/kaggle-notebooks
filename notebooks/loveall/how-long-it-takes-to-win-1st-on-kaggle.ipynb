<center> <h1>How Long It Takes to Win 1st on Kaggle</h1> </center>

I am curious about how long it takes for Kagglers to win their first No. 1 here. I only considered competitions with rewards of USD and jobs since some tutorial competitions may only have a few teams.

The data shows that on average Kagglers take about 1 year to win 1st after registration. However, it seems more and more easier to achieve this compared with earlier days of Kaggle.

```python
%matplotlib inline  

import matplotlib.pyplot as plt
import numpy as np
import plotly.plotly as py
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
sns.set(style="ticks")
```

```python
t = pd.read_csv('../input/Teams.csv')
# t.head()
```

```python
tm =  pd.read_csv('../input/TeamMemberships.csv')
# tm.head()
```

```python
usr =  pd.read_csv('../input/Users.csv')
# usr.head()
```

```python
cp =  pd.read_csv('../input/Competitions.csv')
# cp.head()
```

### Solo or Team Leader

```python
from datetime import datetime

d = dict()

t1 = t[t.Ranking == 1]

for index, row in t1.iterrows():
    cpRow = cp[cp.Id == row.CompetitionId].iloc[0]
    if cpRow.RewardTypeId in [1, 5]:  # Only Reward Type of USD or Jobs
        deadline = cpRow.Deadline 
        if row.TeamLeaderId in d:
            if deadline < d[row.TeamLeaderId]:
                d[row.TeamLeaderId] = deadline
        else:
            d[row.TeamLeaderId] = deadline

df = pd.DataFrame(columns=['Id', 'RegisterDate', 'Duration'])
for k in d:
    reg = datetime.strptime(usr[usr.Id == k].iloc[0].RegisterDate, "%Y-%m-%d %H:%M:%S")
    duration = (datetime.strptime(d[k], "%Y-%m-%d %H:%M:%S") - reg).days
    df = df.append(pd.Series([k, usr[usr.Id == k].iloc[0].RegisterDate, duration], index=['Id', 'RegisterDate', 'Duration']), ignore_index=True)

    
df.describe()
```

```python
# df
df.head()
```

```python
plt.hist(df.Duration)
plt.title("Days Needed to Win 1st")
plt.xlabel("Days Needed")
plt.ylabel("Frequency")

fig = plt.gcf()
```

```python
df["RegisterDate"] = df["RegisterDate"].astype('datetime64[ns]')

ax = df.plot(x = "RegisterDate", y = "Duration", style=['o'], title = "Days Needed to Win 1st")
```

### Discussion:
    
* Only competitions with USD or jobs rewards are considered, since there are tutorial competitions with just a few teams.

* On average, Kagglers takes 483 days (~ 16 months) from registration to win the 1st by solo or team leader.

* It is very impressive that Kaggler 1216 achieved this just in one day.

* It is getting easier to win 1st recently compare to earlier days. I think the reason is Kagglers are so generous to share source code and interviews, which makes it easier to get start.

### Solo, Team Leader or Team Member

```python
d = dict()

t1 = t[t.Ranking == 1]

for index, row in t1.iterrows():
    cpRow = cp[cp.Id == row.CompetitionId].iloc[0]
    if cpRow.RewardTypeId in [1, 5]:  # Only Reward Type of USD or Jobs
        deadline = cpRow.Deadline 
        
        all_members = tm[tm.TeamId == row.Id]
        
        for _, row_member in all_members.iterrows():
            if row_member.UserId in d:
                if deadline < d[row_member.UserId]:
                    d[row_member.UserId] = deadline
            else:
                d[row_member.UserId] = deadline

        

df = pd.DataFrame(columns=['Id', 'RegisterDate', 'Duration'])
for k in d:
    try:
        reg = datetime.strptime(usr[usr.Id == k].iloc[0].RegisterDate, "%Y-%m-%d %H:%M:%S")
        duration = (datetime.strptime(d[k], "%Y-%m-%d %H:%M:%S") - reg).days
        df = df.append(pd.Series([k, usr[usr.Id == k].iloc[0].RegisterDate, duration], index=['Id', 'RegisterDate', 'Duration']), ignore_index=True)
    except:
        pass
    
df.describe()
```

```python
# df
df.head()
```

```python
plt.hist(df.Duration)
plt.title("Days Needed to Win 1st")
plt.xlabel("Days Needed")
plt.ylabel("Frequency")

fig = plt.gcf()
```

```python
df["RegisterDate"] = df["RegisterDate"].astype('datetime64[ns]')

ax = df.plot(x = "RegisterDate", y = "Duration", style=['o'], title = "Days Needed to Win 1st")
```
