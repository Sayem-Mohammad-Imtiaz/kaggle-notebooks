**Selecting employees above the age of 50 and who smokes. Also, what is their average expenditure on health management?**

**Classify the employees into 3 categories.
Category 1 : employees below 18.5 bmi - underweight 
Category 2 : employees with bmi in the range of 18.5 to 24.9 - fit 
Category 3 : employee with bmi over 24.9 - overweight**

**What is the average expenditure of employees in each category. Display the bar chart for the same**

```python
bmi1 = []

for row in df1['bmi']:
    if row < 18.5 :
        bmi1.append('underweight')
    elif row <24.9:
        bmi1.append('fit')
    else:
        bmi1.append('overweight')
df1['Emp_BMI']=bmi1
print(df1)
```

```python
plot1=df1.groupby(["Emp_BMI"])[["charges"]].mean()
plot1
```

**Bar chart of fit/ overweight/ underweight.**

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
x = df1['Emp_BMI']
y = df1['charges']
ax.bar(x,y)
plt.show()
```
