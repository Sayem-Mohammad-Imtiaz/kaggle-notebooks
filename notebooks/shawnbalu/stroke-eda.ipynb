```python
Categorical_Variables = ['gender','hypertension','heart_disease','ever_married','work_type','Residence_type','smoking_status','stroke']
Continous_Variables = ['age','avg_glucose_level','bmi']
```

```python
from matplotlib import pyplot as plt

figure = plt.figure(figsize=(16,16))
i = 431
for col in Categorical_Variables:
    plt.subplot(i)
    df[col].value_counts().plot(kind='bar')
    plt.title(col)
    i = i +1 
plt.show()
```

# Observation of whole dataset
* Female, Non Hypertension people, Non Heart Disease People, Married People, Private working people, Never Smoked people are higher in the dataset.
* Urban and Rural data are almost balanced in the dataset.
* Dataset is highly imbalanced and the stroke affected people are very much low.

```python
from matplotlib import pyplot as plt

figure = plt.figure(figsize=(16,16))
i = 431
for col in Categorical_Variables:
    plt.subplot(i)
    df[df['stroke']==1][col].value_counts().plot(kind='bar')
    plt.title(col)
    i = i +1 
plt.show()
```

# Observation of Stroke People
1. Female are affected more when compared to Men in the stroke.
2. Whoever not having Hypertension and Heart Disease are chances of affecting in the stroke.
3. Never smoked, Married and Private work type people are having high chance in affecting the stroke.
4. Most propbaply in both Rural and Urban people are affected in the stroke.

```python
import seaborn as sns

fig = plt.figure(figsize=(16,16))
i = 431
for col in Continous_Variables:
    plt.subplot(i)
    sns.kdeplot(df[col], shade=True)
    i = i +1
for col in Continous_Variables:
    plt.subplot(i)
    sns.histplot(df[col], kde=True)
    i = i +1

plt.show()
```

# Observation of Age, Glucose level and BMI
1. All age of people were available in the dataset

3. 15 to 50 bmi were high in the dataset.

```python
import seaborn as sns

fig = plt.figure(figsize=(16,16))
i = 431
for col in Continous_Variables:
    plt.subplot(i)
    sns.kdeplot(df[df['stroke'] == 1][col], shade=True)
    i = i +1
for col in Continous_Variables:
    plt.subplot(i)
    sns.histplot(df[df['stroke'] == 1][col], kde=True)
    i = i +1
plt.show()
```
