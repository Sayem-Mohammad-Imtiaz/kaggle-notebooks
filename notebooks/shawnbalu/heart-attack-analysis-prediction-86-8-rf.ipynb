```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
y = df.pop('output')
x = df
x_train,x_test,y_train,y_test = train_test_split(x, y,test_size=0.25, random_state=100)
```

```python
print(x_train.columns)
```

```python
model = RandomForestClassifier().fit(x_train,y_train)
```

```python
model.feature_importances_
```

```python
plt.barh(x_train.columns, model.feature_importances_)
plt.show()
```

```python
sorted_indx = model.feature_importances_.argsort()
```

```python
plt.barh(x_train.columns[sorted_indx], model.feature_importances_[sorted_indx])
plt.show()
```

```python
df_new = df[['oldpeak','caa','thalachh','cp','thall','chol','age','trtbps']]
```

```python
print(df_new.head())
```

```python
x_train,x_test,y_train,y_test = train_test_split(df_new,y,test_size=0.25,random_state=100)
model_new = RandomForestClassifier(n_estimators=500, max_features=2, max_depth=5, class_weight='balanced')
model_new.fit(x_train,y_train)
```

```python
y_train.value_counts()
```

```python
y_predicted = model_new.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy_score(y_test,y_predicted)
```

```python
classification_report(y_test,y_predicted)
```

Getting the Accuracy score of **84.2 %**
