```python
df1 = pd.read_csv('../input/ipl-data-set/deliveries.csv')
df1.head()
```

```python
df1.columns.unique()
```

```python
df1.batting_team.unique()
```

```python
df1.batting_team.replace({"Rising Pune Supergiant":"Rising Pune Supergiants"},inplace = True)
df1.bowling_team.replace({"Rising Pune Supergiant":"Rising Pune Supergiants"},inplace = True)
```

**Venues that hosted the most no of matches**

```python
city_counts=df.groupby('city').apply(lambda x:x['city'].count()).reset_index(name='Match Counts')
req_values = city_counts.sort_values('Match Counts' , ascending = False).reset_index(drop = True)
req_values = req_values[:20]
```

```python
plt.rcParams["figure.figsize"] = (20, 10)
plt.title('Venues that hosted the most number of matches', size =30)
plt.xlabel('Venues',size =25)
plt.ylabel('Number of matches',size =25)
plt.xticks(rotation=90,size = 15)
plt.yticks(size = 15)
count = 0
for i in req_values['Match Counts']:
    plt.text(count,i+1,str(i),color='black',size=15)
    count+=1
plt.bar(req_values['city'],req_values['Match Counts'])
plt.show()
```

**Top 15 IPL Innings**

```python
a = df1.groupby(['match_id','batsman'])['batsman_runs'].sum().reset_index(name ="Innings_Runs")
b = a.sort_values(by = 'Innings_Runs',ascending = False)
c= b.reset_index(drop = True)
c = c[:15]
print("Top 15 best innings of the IPL:\n")
print(c)
```

**Best Fielder**

```python
# Best Fielder
bf = df1.groupby('fielder').apply(lambda x : x).reset_index()
bf = df1.groupby('fielder').count() 
bf = bf.dismissal_kind.reset_index(name='Dismissals')
bf = bf.sort_values(by='Dismissals',ascending=False)
bf = bf[0:10]

plt.title("Best IPL Fielders")
plt.bar(bf.fielder , bf.Dismissals)
plt.xlabel("Best Fielders")
plt.ylabel("No of  Dismissals")
count = 0
for i in bf.Dismissals:
    plt.text(count-0.2,i-4,str(i))
    count+=1
plt.xticks(rotation = 90)
plt.yticks()
plt.show()
```

**Most Number of Wins by a team**

```python
most_no_of_wins = df.groupby('winner').apply(lambda x: x).reset_index()
most_no_of_wins = most_no_of_wins.groupby('winner').count()
most_no_of_wins = most_no_of_wins.city.reset_index(name='No_Of_Wins')
most_no_of_wins = most_no_of_wins.sort_values(by='No_Of_Wins',ascending=False)

plt.title("Most Number Of Wins")
plt.bar(most_no_of_wins.winner , most_no_of_wins.No_Of_Wins)
plt.xlabel("Teams")
plt.ylabel("No of  Wins")
count = 0
for i in most_no_of_wins.No_Of_Wins:
    plt.text(count-0.2,i-4,str(i))
    count+=1
plt.xticks(rotation = 90)
plt.yticks()
plt.show()
```

**Most MOM Awards**

```python
mom = df.groupby('player_of_match').apply(lambda x: x).reset_index()
mom = mom.groupby('player_of_match').count()
mom = mom.city.reset_index(name='No_Of_Awards')
mom = mom.sort_values(by='No_Of_Awards',ascending=False)
mom = mom.head(10)

plt.title("Most Number Of MOM Awards")
plt.bar(mom.player_of_match , mom.No_Of_Awards)
plt.xlabel("Name")
plt.ylabel("No of  Awards")
count = 0
for i in mom.No_Of_Awards:
    plt.text(count-0.2,i,str(i))
    count+=1
plt.xticks(rotation = 90)
plt.yticks()
plt.show()
```

**Teamwise Bowling Average**

```python
fis = df1[df1['inning'] == 1]
fis = fis.groupby(['bowling_team','match_id']).apply(lambda x : np.sum(x['total_runs'])).reset_index(name = 'Innings_Total')
fisa = fis.groupby(['bowling_team']).apply(lambda x : np.mean(x['Innings_Total'])).reset_index(name = 'Innings_Average')
plt.plot(fisa.bowling_team,fisa.Innings_Average,'b')
sis = df1[df1['inning'] == 2]
sis = sis.groupby(['bowling_team','match_id']).apply(lambda x : np.sum(x['total_runs'])).reset_index(name = 'Innings_Total')
sisa = sis.groupby(['bowling_team']).apply(lambda x : np.mean(x['Innings_Total'])).reset_index(name = 'Innings_Average')
plt.plot(fisa.bowling_team,fisa.Innings_Average,'b')
plt.plot(sisa.bowling_team,sisa.Innings_Average,'r')
plt.title('TEAMWISE BOWLING AVERAGE')
plt.xticks(rotation = 90)
```

**Toss Decisions:**

**Matches hosted in each city**

```python
venue = df.groupby('venue').apply(lambda x: x).reset_index()
venue = venue.groupby('venue').count()
venue = venue.city.reset_index(name='No_Of_Matches')
venue = venue.sort_values(by='No_Of_Matches').reset_index(drop = True)
print(venue)

venue = venue.sort_values('No_Of_Matches', ascending = False)
venue = venue[:10]
plt.bar(venue.venue,venue.No_Of_Matches)
plt.title("TOP 10 Venues" , size = 20)
plt.xlabel('Venues', size = 15)
plt.ylabel('No Of Matches' , size = 15)
plt.xticks(rotation = 90)
plt.yticks()
```
