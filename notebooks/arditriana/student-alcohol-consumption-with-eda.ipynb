### **Univariant Analysis**

```python
f = plt.figure(figsize=(17,5))

ax = f.add_subplot(121)
sns.distplot(df.age,color='orange')
ax.set_title('Distribution of age people')

ax=f.add_subplot(122)
sns.distplot(df.studytime,color='blue')
ax.set_title('Distribution of study time in people')
```

### **Bivariant Analysis**

```python
f = plt.figure(figsize=(17,5))

ax = f.add_subplot(121)
sns.countplot(x="sex", hue ="Dalc", data=df)

ax=f.add_subplot(122)
sns.countplot(x="sex", hue ="Walc", data=df)
```

```python
sns.regplot(x=df.traveltime, y=df.health)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True)
plt.show()
```
