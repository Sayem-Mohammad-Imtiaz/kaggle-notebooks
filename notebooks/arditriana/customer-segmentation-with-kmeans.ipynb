```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/mall-customers/Mall_Customers.csv')
df.head()
```

```python
# menghapus kolom
df = df.drop(columns=(['CustomerID', 'Genre']))
```

```python
# mengubah nama kolom
df = df.rename(columns={
    'Age' : 'age', 
    'Annual Income (k$)': 'annual_income',
    'Spending Score (1-100)': 'spending_score'
})
```

```python
X = df
X
```

```python
from sklearn.cluster import KMeans
```

```python
cluster = []
for k in range (1, 11):
    kmean = KMeans(n_clusters=k).fit(X)
    cluster.append(kmean.inertia_)
```

```python
plt.figure(figsize=(12, 8))
plt.plot(range(1, 11), cluster, 'r-')
plt.xlabel('Inertia')
plt.ylabel('n_cluster')
plt.show()
```

```python
km = KMeans(n_clusters=5).fit(X)

# menambah kolom baru pada datasets
X['Labels'] = km.labels_

plt.figure(figsize=(12, 8))
sns.scatterplot(X['annual_income'], X['spending_score'], hue=X['Labels'], palette=sns.color_palette('hls', 5))
plt.show()
```
