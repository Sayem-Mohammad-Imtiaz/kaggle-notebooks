```python
df_DF5 = df_games_details[df_games_details['PLAYER_NAME'] == "De'Aaron Fox"]
```

```python
df_DF5.loc[(df_DF5.GAME_ID >= 22000079), 'SEASON'] = '20-21'
```

```python
def category_season(x):
    if x>=22000079:
        return '20-21'
    
    elif x>=21900119:
        return '19-20'
    
    elif x>=21800197:
        return '18-19'
    
    elif x>=21700013:
        return '17-18'
```

```python
df_DF5['SEASON'] = df_DF5.GAME_ID.apply(category_season)
```
