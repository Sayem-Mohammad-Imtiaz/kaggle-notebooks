# Analysis of Books by Year

```python
years = sorted(data['Year'].unique())
year_group = data.groupby('Year')
# alltime dataset only keep the data of first year for those books have records more than once 
```

```python
year_group.Genre.count()
# year_group.Genre.value_counts()
```

```python
num_of_book = []
non_fiction = []
fiction = []
genre = []
for year in years:
    yearbook = year_group.get_group(year)['Genre'].value_counts()
    num_of_book.append(yearbook.sum())
    genre.append(yearbook)
    non_fiction.append(yearbook[0])
    fiction.append(yearbook[1])
```

```python
fiction_rate = []
for i in range(0,11):
    fiction_rate.append(fiction[i]/num_of_book[i])
non_fiction_rate = []
for i in range(0,11):
    non_fiction_rate.append(non_fiction[i]/num_of_book[i])
```

```python
plt.bar(years, num_of_book)
plt.xlabel('Year')
plt.ylabel('Number of Books')
```

```python
sns.lineplot(years, fiction_rate)
plt.ylim(bottom=0)
plt.xlabel('Year')
plt.ylabel('Rate of Genre Fiction')
```

```python
sns.lineplot(years, non_fiction_rate)
plt.ylim(bottom=0)
plt.xlabel('Year')
plt.ylabel('Rate of Genre Non-Fiction')
```

# Top 10 Books with Most Reviews 

```python
top10 = dict()
for year in years:
    data_year = year_group.get_group(year)
    top10[year] = data_year.sort_values('Reviews',ascending=False)[:10]
```
