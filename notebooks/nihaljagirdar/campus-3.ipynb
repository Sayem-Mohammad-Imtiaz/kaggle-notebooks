```python
data=pd.read_csv('/kaggle/input/factors-affecting-campus-placement/Placement_Data_Full_Class.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data.drop(columns=['sl_no'],inplace=True)
data.head()
```

```python
data.salary=data.salary.fillna(0)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
x=data.drop(columns=['salary'])

y=data[['salary']]
```

```python
nom_cols=['gender','hsc_s','degree_t','specialisation']
ord_cols=['ssc_b','hsc_b','workex','status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
trans_data=make_column_transformer((OneHotEncoder(),[0,5,7,10]),(OrdinalEncoder(),[2,4,8,12]),remainder='passthrough')
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
from sklearn.pipeline import make_pipeline
from sklearn import set_config
set_config(display='diagram')
pipe=make_pipeline(trans_data,model)
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(pred,y_test)
```
