This is the "**Iris**" dataset. Originally published at UCI Machine Learning Repository: Iris Data Set, this small dataset from 1936 is often used for testing out machine learning algorithms and visualizations.Each row of the table represents an iris flower, including its species and dimensions of its botanical parts, sepal and petal, in centimeters.

* ****First load the iris dataset from the csv file by using read_csv method of pandas library.****

```python
import pandas as pd
df = pd.read_csv("../input/iris-flower-dataset/IRIS.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

* **By using LabelEncoder technique convert the categorical values into numbers.And drop the species(categorical data) column.**

```python
from sklearn.preprocessing import LabelEncoder
enc = LabelEncoder()
df['target'] = enc.fit_transform(df['species'])
```

```python
df.drop(['species'],axis='columns')
```

```python
import matplotlib.pyplot as plt
plt.legend('Flower species')
plt.xlabel('sepal length(cm)')
plt.ylabel('output')
plt.scatter(df['sepal_length'],df['target'])
plt.show()
```

* **Split the model into training and testing by using train_test_split function which is present in the sklearn.model_selection'**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,0:4].values,df['target'],test_size=0.2)
```

***Train the model by using algorithm Random Forest*.A random forest is a meta estimator that fits a number of decision tree classifiers on various sub-samples of the dataset and uses averaging to improve the predictive accuracy and control over-fitting.**

```python
from sklearn.ensemble import RandomForestClassifier
RFC = RandomForestClassifier()
RFC.fit(x_train,y_train)
```

**Predict the output from the trined module.**

```python
y_predict = RFC.predict(x_test)
y_predict
```

**By using confusion matrix measure the performance of the given dataset.**

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,classification_report,precision_score
cm = confusion_matrix(y_test,y_predict)
cm
```

```python
print(accuracy_score(y_test,y_predict))
```
