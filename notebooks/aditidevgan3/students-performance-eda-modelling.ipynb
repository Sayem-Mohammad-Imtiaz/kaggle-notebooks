```python
fig, axes = plt.subplots(figsize=(12,6))
sns.countplot(data=df, x="parental level of education", hue="test preparation course", palette=['#F3CEE8','#B7D1F8'])
sns.set(style="whitegrid")
axes.legend(loc='upper right',frameon=True)
axes.set_ylim(0, 160)
plt.title("*! Test preparation according to parental level of education !*")
plt.show()
print("*! On the basis of the plot we can state that the high school pass out parent's students have shown the better results. !*")
```

```python
sns.catplot(x="lunch", y="math score", hue="gender", kind="point", data=df, palette="ch:s=-.2,r=.6")
plt.show()
```

```python
a = df['math score'].sum()
b = df['reading score'].sum()
c = df['writing score'].sum()

# Data to plot
fig, axes = plt.subplots(figsize=(12,6))
labels = 'math score', 'reading score', 'writing score'
sizes = a,b,c
colors = ['gold', 'yellowgreen', 'lightcoral']
explode = (0.1, 0, 0) 

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=180)

plt.axis('equal')
plt.show()
```

```python
sns.violinplot(x=df.gender, y=df['reading score'], palette="viridis", inner="stick")
plt.show()
```

```python
sns.pairplot(df, hue = "race/ethnicity", palette="Spectral")
plt.show()
```

# Modelling üìäüë®‚Äçüíªüìâ

## Random Forest 

```python
X, Y = make_classification(n_samples=1000, n_features=14, n_informative=9, n_redundant=5, random_state=3)
print(X.shape)
print(Y.shape)
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
#Import Random Forest Model
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,2),"%")
```

## Support Vector Machines 

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",round(metrics.accuracy_score(y_test, y_pred)*100,2),"%")
```
