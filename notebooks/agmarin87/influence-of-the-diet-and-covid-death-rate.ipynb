I have found this datasheet quite interesting. I would like to know if diet is important to avoid getting the COVID. For this aim, I will the datasheet with pandas and plot the correlation map with the seaborn library.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb #for the correlation heatmap
start_dir = r'D:\TRABAJO\Mis_programas\COVID food influence' 

df = pd.read_csv('D:\TRABAJO\Mis_programas\COVID food influence\Fat_Supply_Quantity_Data.csv', header=0)
#Correlation plot
plot1 = plt.figure(1)
plt.title('COVID Meat correlation')
df_pearson = df.iloc[:,[2,3,6,7,8,11,12,24,26,27,28]] 
pearsoncorr = df_pearson.corr(method='pearson')
sb.heatmap(pearsoncorr, 
            xticklabels=pearsoncorr.columns,
            yticklabels=pearsoncorr.columns,
            annot=True,
            annot_kws={"fontsize":8},
            linewidth=0.3)
#watermark
plt.text(-1, -1.5, 'By Antonio García',
         fontsize=20, color='gray',
         ha='center', va='center', alpha=0.8)
#save plot
plt.savefig(r'D:\TRABAJO\Mis_programas\COVID food influence\COVID-meat correlation.png', 
            dpi=100, bbox_inches='tight')

#Correlation plot
plot1 = plt.figure(2)
plt.title('COVID Vegetables correlation')
df_pearson = df.iloc[:,[5,13,14,15,16,20,21,22,24,26,27,28]] 
pearsoncorr = df_pearson.corr(method='pearson')
sb.heatmap(pearsoncorr, 
            xticklabels=pearsoncorr.columns,
            yticklabels=pearsoncorr.columns,
            annot=True,
            annot_kws={"fontsize":8},
            linewidth=0.3)
#watermark
plt.text(-1, -1.5, 'By Antonio García',
         fontsize=20, color='gray',
         ha='center', va='center', alpha=0.8)
```

It looks like there is not a strong correlation between the consumption of meat products and the death rate (value are not higher than 0.5). However, it is possible to see that a vegetarian diet might be better due to the negative correlation values. And it also helps people to avoid obesity.

Also, I would like to know, which countries have a higher degree of obesity. For this, I am plotting the following map with the following code.

```python
#WORLD MAP
import plotly.express as px
import pandas as pd
df = pd.read_csv("../input/covid19-healthy-diet-dataset/Fat_Supply_Quantity_Data.csv", header=0)
fig = px.choropleth(df, locations="Country",
                    color="Obesity", 
                    locationmode='country names',
                    color_continuous_scale="Inferno",
                    title="Obesity in the world",
                    height=600)
fig.show()
```
