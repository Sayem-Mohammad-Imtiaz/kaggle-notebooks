```python
data['MinTemp'].fillna(data['MinTemp'].mean(),inplace=True)
data['MaxTemp'].fillna(data['MaxTemp'].mean(),inplace=True)
data['Rainfall'].fillna(data['Rainfall'].mean(),inplace=True)
data['Evaporation'].fillna(data['Evaporation'].mean(),inplace=True)
data['Sunshine'].fillna(data['Sunshine'].mean(),inplace=True)
data['WindGustSpeed'].fillna(data['WindGustSpeed'].mean(),inplace=True)
data['WindSpeed9am'].fillna(data['WindSpeed9am'].mean(),inplace=True)
data['WindSpeed3pm'].fillna(data['WindSpeed3pm'].mean(),inplace=True)
data['Humidity9am'].fillna(data['Humidity9am'].mean(),inplace=True)
data['Humidity3pm'].fillna(data['Humidity3pm'].mean(),inplace=True)
data['Pressure9am'].fillna(data['Pressure9am'].mean(),inplace=True)
data['Pressure3pm'].fillna(data['Pressure3pm'].mean(),inplace=True)
data['Cloud9am'].fillna(data['Cloud9am'].mean(),inplace=True)
data['Cloud3pm'].fillna(data['Cloud3pm'].mean(),inplace=True)
data['Temp9am'].fillna(data['Temp9am'].mean(),inplace=True)
data['Temp3pm'].fillna(data['Temp3pm'].mean(),inplace=True)
data['WindGustDir'].fillna(data['WindGustDir'].value_counts().index[0],inplace=True)
#data['WindGustDir']=label_encoder.fit_transform(data['WindGustDir'])
data['WindDir9am'].fillna(data['WindDir9am'].value_counts().index[0],inplace=True)
#data['WindDir9am']=label_encoder.fit_transform(data['WindDir9am'])
data['WindDir3pm'].fillna(data['WindDir3pm'].value_counts().index[0],inplace=True)
#data['WindDir3pm']=label_encoder.fit_transform(data['WindDir3pm'])
data['RainToday'].fillna(data['RainToday'].value_counts().index[0],inplace=True)
#data['RainToday']=label_encoder.fit_transform(data['RainToday'])
data['RainTomorrow'].fillna(data['RainTomorrow'].value_counts().index[0],inplace=True)
#data['RainTomorrow']=label_encoder.fit_transform(data['RainTomorrow'])

data.isna().sum()
```

```python
data.head()
```

```python
data['WindGustDir']=label_encoder.fit_transform(data['WindGustDir'])
data['WindDir9am']=label_encoder.fit_transform(data['WindDir9am'])
data['WindDir3pm']=label_encoder.fit_transform(data['WindDir3pm'])
data['RainToday']=label_encoder.fit_transform(data['RainToday'])
data['RainTomorrow']=label_encoder.fit_transform(data['RainTomorrow'])
data['Location']=label_encoder.fit_transform(data['Location'])
```

```python
x=data[['Location', 'MinTemp', 'MaxTemp', 'Rainfall', 'Evaporation',
       'Sunshine', 'WindGustDir', 'WindGustSpeed', 'WindDir9am', 'WindDir3pm',
       'WindSpeed9am', 'WindSpeed3pm', 'Humidity9am', 'Humidity3pm',
       'Pressure9am', 'Pressure3pm', 'Cloud9am', 'Cloud3pm', 'Temp9am',
       'Temp3pm', 'RainToday']]
y=data['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
model=RandomForestClassifier(n_estimators=200)
#model=xgb.XGBClassifier(n_estimators=200)
```

```python
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,mean_squared_error
```

```python
accuracy_score(y_test,prediction)
```

```python
confusion_matrix=pd.crosstab(y_test,prediction,rownames=['actual'],colnames=['predicted'])
sns.heatmap(confusion_matrix,annot=True)
```

```python
mse=mean_squared_error(y_test,prediction)
mse
```

```python
rmse=np.sqrt(mse)
rmse
```
