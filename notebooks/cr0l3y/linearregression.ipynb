```python
#transformando os dados

#transformando as variaveis binárias
df.loc[df['school'] == 'GP', 'school'] = 1
df.loc[df['school'] == 'MS', 'school'] = 0

df.loc[df['sex'] == 'M', 'sex'] = 1
df.loc[df['sex'] == 'F', 'sex'] = 0

df.loc[df['address'] == 'U', 'address'] = 1
df.loc[df['address'] == 'R', 'address'] = 0

df.loc[df['famsize'] == 'GT3', 'famsize'] = 1
df.loc[df['famsize'] == 'LE3', 'famsize'] = 0


df.loc[df['Pstatus'] == 'T', 'Pstatus'] = 1
df.loc[df['Pstatus'] == 'A', 'Pstatus'] = 0

```

```python
#transformando as variaveis yes/no

for col in ['schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher', 'internet', 'romantic']:

    df[col] = df[col].map({'yes': 1, 'no': 0})
```

```python
#transformando as variaveis categoricas usando o categoru do pandas

for col in ['Mjob','Fjob', 'reason', 'guardian']:
    
    #transformando o texto em categoria
    df[col] = df[col].astype('category')
    
    
    #salvando apenas os numeros de cada categoria
    df[col] = df[col].cat.codes
```

```python
df.info()
```

```python
#Vamos criar os datasets de treino e teste


#Importando o train_test_split do scikit learn

from sklearn.model_selection import train_test_split

#separando o dataframe em treino e teste e usando uma semente aleatoria
#parar de reproduzir os dados
#por padrao test_size é 0.25

train, test = train_test_split(df, test_size=0.2, random_state=42)



```

```python
train.shape, test.shape
```

```python
#selecionando as colunas que serao usadas para treino

#nao vamos usar as notas intermediarias para treinamento
# e tambem nao usaerrmos a variavel target 'G3'

remove = ['G1', 'G2', 'G3']

# Lista com as principais colunas a serem usadas
feats = [col for col in df.columns if col not in remove]
```

```python
feats
```

```python
#criando o modelo de regreção
from sklearn import linear_model

regr = linear_model.LinearRegression()
```

```python
#treinando o modelo
regr.fit(X = train[feats], y= train['G3'])
```

```python
#gerando predições com base no modelo treinado

preds = regr.predict(X = test[feats])
```

```python
preds
```

```python
test['G3'].head()
```

```python
#avaliando o modelo de acordo com o mean squared error

from sklearn.metrics import mean_squared_error


#chama a função passando os valores reais e os calores preditos
mean_squared_error(test['G3'], preds)


```

```python
#avaliabdi i nidek oir neui de r2 Score
from sklearn.metrics import r2_score

r2_score(test['G3'], preds)
```

```python
#Qual o desempenho do modelo do modelo para os dados/variaveis de treino?


#Gerando previsões para os dados de treino usando o modelo treinado

preds2= regr.predict(X=train[feats])


#Analisando as metricas
mean_squared_error(train['G3'], preds2), r2_score(train['G3'], preds2)

```
