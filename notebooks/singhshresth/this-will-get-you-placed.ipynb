```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
import plotly.express as px
import plotly.graph_objects as go
data = pd.read_csv('/kaggle/input/factors-affecting-campus-placement/Placement_Data_Full_Class.csv')
```

**<span style="color:#8B4C39;font-size:25px"> Number of Students having Work Experience </span>**

```python
import seaborn as sns
sns.countplot(data.workex)
```

**65%** people had no work experience during the time of placements. Does this affect their salary? more on this later.

**<span style="color:#8B4C39;font-size:25px"> Gender Distribution </span>**

```python
sns.countplot(data.gender)
```

There's a huge **imbalance** between both the genders that appeared for placements by **men** outnumbering **women** by a wide margin

**<span style="color:#8B4C39;font-size:25px"> Placement Scenario </span>**

```python
fig = go.Figure([go.Pie(labels=data['status'].unique(), values=data['status'].value_counts())])
fig.show()
```

The placements in the campus are **average** with almost **69%** ;) chances of getting placed. It would boil down on your competency

**<span style="color:#8B4C39;font-size:25px"> Undergraduate Stream Distribution </span>**

```python
fig = go.Figure([go.Pie(labels=data['degree_t'].unique(), values=data['degree_t'].value_counts())])
fig.show()
```

**<span style="color:#8B4C39;font-size:25px"> Does your gender increase the chance of you getting placed? </span>**

```python
data.groupby('gender')['status'].value_counts()
```

If you calculate the ratio to find out whether your gender increases your chance of getting placed. The answer would be **no**, so we can infer that the procedure is carried out in the **absence of sexual bias**.

**<span style="color:#8B4C39;font-size:25px"> Do Senior Secondary Grades affect your Placement? </span>**

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='status',y='ssc_p',data=data)
```

Unfortunately your secondary education marks affects your placement process but the threshold is set to about 60-65%. This is evident since some companies filter out some candidates on their 10th and 12th percentage. 

**<span style="color:#8B4C39;font-size:25px"> Do High School Grades affect your Placement? </span>**

```python
plt.figure(figsize=(10,7))
sns.boxplot(x='status',y='hsc_p',data=data)
```

We observe a similar trend in high school percentage. Anything above 65% can be considered a safe zone in both the standards

**<span style="color:#8B4C39;font-size:25px"> High School Percentage Distribution on basis of Stream </span>**

```python
commerce = data[data['hsc_s']=='Commerce']['hsc_p']
science = data[data['hsc_s']=='Science']['hsc_p']
arts = data[data['hsc_s']=='Arts']['hsc_p']

plt.figure(figsize=(10,5))
sns.distplot(commerce, color="skyblue", label="Commerce",hist=False)
sns.distplot(science, color="red", label="Science",hist=False)
sns.distplot(arts, color="green", label="Arts",hist=False)
```

**Commerce** students tend to get the **best marks** in our population sample

**<span style="color:#8B4C39;font-size:25px"> Undergraduate Degree Percentage Distribution on basis of Stream </span>**

```python
tech = data[data['degree_t']=='Sci&Tech']['degree_p']
com_mgt = data[data['degree_t']=='Comm&Mgmt']['degree_p']
others = data[data['degree_t']=='Others']['degree_p']

plt.figure(figsize=(10,5))
sns.distplot(tech, color="skyblue", label="Science & Tech",hist=False)
sns.distplot(com_mgt, color="red", label="Commerce & Management",hist=False)
sns.distplot(others, color="green", label="Others",hist=False)
```

People with a **technological degree** tend to get **better percentage** than the rest of the two. Thank God I chose engineering ;)

**<span style="color:#8B4C39;font-size:25px"> Does Undergraduate College Percentage Determine your Salary? </span>**

```python
plt.figure(figsize=(12,5))
sns.scatterplot(y='degree_p',x='salary',data=data)
```

The person who scored **65%** got the highest package in the college, the second highest package has score **below 60%**. We see the trend that people having relatively exceptional salary packages didn't stand out when it came to scoring marks. **Skills matter more than your grades**.

**<span style="color:#8B4C39;font-size:25px"> Does Masters Degree Percentage Determine your Salary? </span>**

```python
plt.figure(figsize=(12,5))
sns.scatterplot(y='mba_p',x='salary',data=data)
```

Anything above **60%** is good enough to get a decent package when you pursue Master of Business Administration in this college. We also observe that people pursuing Masters generally score **lower percentage** than undergraduates.

**<span style="color:#8B4C39;font-size:20px"> Does your Degree Percentage Influence your Employability Test Percentage? Does Work Experience have an affect?  </span>**

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='etest_p',y='degree_p',hue='workex',data=data)
```

You can raise the argument that work experience does not have a reasonable affect on employability test percentage since people with work experience are evenly spread on the x-axis. One may rightfully argue and doubt the procedure for calculating the employability percentage since it makes sense that people having work experience would have higher employability percentage which is not evident in the data. **This needs to be investigated further**

**<span style="color:#8B4C39;font-size:20px"> Is Employability Test Percentage even a Useful Metric to Evaluate your Skill ?  </span>**

```python
plt.figure(figsize=(12,5))
sns.scatterplot(y='etest_p',x='salary',data=data)
```

Like I pointed out above the employability test procedure is not reliant and needs to be re-evaluated. You can see from the data that people having **high employability test percentage (between 90-100)** couldn't manage to grab a package above **450k**. The person who grabbed the highest package didn't score **85%** and the second highest package didn't even cross **60%**. Ironically most people scoring above **90%** didn't cross **300k**.

**<span style="color:#8B4C39;font-size:20px"> How Does your Degree Percentage and Work Experience Influence your Salary </span>**

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='salary',y='degree_p',hue='workex',data=data)
```

This is a very interesting visualization. The People who have work experience really stand out in terms of the salary package they bag but the catch is they do not have exceptional degree percentage. It's average at best. The conclusion is that **your degree percentage won't help a lot in getting you a high package if that's your end goal**

**<span style="color:#8B4C39;font-size:25px">Does your Specialisation Affect your Salary? </span>**

```python
plt.figure(figsize=(12,5))
sns.scatterplot(x='salary',y='mba_p',hue='specialisation',data=data)
```

**In the specialisation aspect Masters in Marketing and Finance would pay you better than Marketing and HR.**
