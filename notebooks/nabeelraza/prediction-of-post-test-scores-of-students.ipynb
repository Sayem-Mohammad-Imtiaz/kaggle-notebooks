```python
data = pd.read_csv("/kaggle/input/predict-test-scores-of-students/test_scores.csv")
data.head()
```

```python
data.shape
```

```python
features = ['school', 'school_setting', 'school_type',
            'teaching_method', 'n_student', 'gender', 'pretest', 'lunch']
X = data[features]
y = data['posttest']
```

```python
data.isna().any()# the data is all cleaned up!
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, random_state=0)
```

```python
cat_variables = [i for i in X.columns if X[i].dtype == object]
cat_variables
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.compose import ColumnTransformer
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, cat_variables)
    ]
)

model = RandomForestRegressor(n_estimators=50, random_state=0)
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', model)
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
predictions = pipeline.predict(X_valid)
predictions
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mae = mean_absolute_error(y_valid, predictions)
mse = mean_squared_error(y_valid, predictions)
print("The MAE is: ", mae)
print("The MSE is: ", mse)
```
