Load the data to the dataet and then begin analyzing

```python
dataset=pd.read_csv('/kaggle/input/health-care-data-set-on-heart-attack-possibility/heart.csv')
dataset.info()
```

The dataset consists of 14 columns. There is nothing missing here and all the data is in numeric format.
Thus, we can start by directly checking the mean, deviation adn such parameters for each columns.

```python
dataset.describe()
```

The following columns are defined categorically:
* Chest pain type(cp)
* resting electrocardiographic result (restecg)
* thal

The first thing to do is to split between X(features) and Y(target value)

```python
X=dataset.drop(columns='target')
Y=dataset['target']
```

```python
from sklearn import model_selection
seed=3
X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,test_size=0.3,random_state=seed)
```

Since, it is a binary classification problem,we would be using Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=3,solver='liblinear',max_iter=400)
lr.fit(X_train, Y_train)
```

```python
Y_pred=lr.predict(X_test)
```

Now, just need to compare between the actual value and the predicted value...

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import accuracy_score
mae=mean_absolute_error(Y_test,Y_pred)
acc=accuracy_score(Y_test,Y_pred)
print(mae)
print(acc)
```
