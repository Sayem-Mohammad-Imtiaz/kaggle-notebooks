- Создайте равномерную сетку из 20 точек на отрезке $\left[0; \frac{\pi}{2}\right]$.
- Проверьте для каждой точки выполнение тригонометрического тождества $\cos\alpha = \sin\left(\frac{\pi}{2}-\alpha\right)$. Для этого: а) создайте массив $\cos\alpha$; б) создайте массив $\sin\left(\frac{\pi}{2}-\alpha\right)$; в) создайте логическую функцию **close(x, y, eps)**, которая возвращает **True**, если $|x-y|<\varepsilon$, и **False** в противном случае; г) примените данную функцию к двум массивам, созданным выше; д) убедитесь, что все значения в полученном массиве равны **True**.

```python
b = np.linspace(0, np.pi/2, 20)
b
```

```python
cos = np.cos(b)
cos
```

```python
sin = np.sin(np.pi/2-b)
sin
```

```python
def close(x, y, eps):
    if np.all(abs(x - y)<eps):
        return True
    else:
        return False
```

```python
close(cos,sin,1e-6)
```

```python
M = np.random.normal(5, 2.5, (10, 5))
M
```

```python
np.mean(M[4,:])
```

```python
np.median(M[:,2])
```

```python
print(np.sum(M, axis=1))
print(np.sum(M, axis=0))
print(np.sum(M))
```

```python
np.linalg.det(M[:3, :3])
```

```python
np.sum(M, axis=1)[np.sum(M, axis=1)<np.mean(np.sum(M, axis=1))].size
```

```python
N=M[M[:,0]>M[:,4],:]
print(N)
N.shape
```

```python
K=M[np.sum(M, axis=1)>np.mean(np.sum(M, axis=1))]
print(K)
np.linalg.matrix_rank(K)
```
