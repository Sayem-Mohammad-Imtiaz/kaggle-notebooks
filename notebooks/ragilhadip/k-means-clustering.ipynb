```python
df = pd.read_csv('../input/customer-segmentation-tutorial-in-python/Mall_Customers.csv')
df.head()
```

```python
features = df.drop(['CustomerID', 'Gender'], axis='columns')
features.head()
```

```python
from sklearn.cluster import KMeans
 
clusters = []
for i in range(1,11):
  kmc = KMeans(n_clusters=i).fit(features)
  clusters.append(kmc.inertia_)
```

```python
clusters
```

```python
import matplotlib.pyplot as plt
import seaborn as sb


fig, ax = plt.subplots(figsize=(8, 4))
sb.lineplot(x=list(range(1, 11)), y=clusters, ax=ax)
ax.set_title('Cari Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
```

```python
kmc5 = KMeans(n_clusters=5).fit(features)
 
features['Labels'] = kmc5.labels_

plt.figure(figsize=(8,4))
sb.scatterplot(features['Annual Income (k$)'], features['Spending Score (1-100)'], hue=features['Labels'],
                palette=sb.color_palette('hls', 5))
plt.title('KMeans With 5 Cluster')
```
