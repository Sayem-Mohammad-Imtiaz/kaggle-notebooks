# Analyses

From here, I will analyze the sata that was loaded as an dataframe object.



```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

selected_coutry_1='Denmark'
selected_coutry_2='Denmark'
df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1 = df1[df1['country']==selected_coutry_1]
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2 = df2[df2['country']==selected_coutry_2]
df1.index = pd.to_datetime(df1['date'])
df2.index = pd.to_datetime(df2['date'])
```

```python
print(df1.columns)
print(df2.columns)
```

```python
print(np.unique(df2['country']))
```

```python
fig,ax = plt.subplots(figsize=(10,5))
ax.plot(df1.index,df1['daily_new_cases'],label='Cases')
ax1 = ax.twinx()
ax1.plot(df2.index,df2['daily_vaccinations'],'r',label='Vaccinations')
ax1.legend()
ax.legend()
plt.show()
```

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_new_cases    = go.Scatter(x=df1.index,y=df1['daily_new_cases'],yaxis='y1',name='Daily confirmed cases')
daily_vaccinations = go.Scatter(x=df2.index,y=df2['daily_vaccinations'],yaxis='y2',name='Daily vaccinations')

layout_obj = go.Layout(title='COVID vs. Vaccines',xaxis=dict(title='Date'),yaxis=dict(title='Daily cases'),yaxis2=dict(title='Vaccinations',side='right',overlaying='y'))
fig = go.Figure(data=[daily_new_cases,daily_vaccinations],layout=layout_obj)
iplot(fig)
fig.write_html('outcome.html')
```

```python
style_object = df1.style.background_gradient(cmap='jet').highlight_max('daily_new_cases').set_caption('Daily cases')
display(style_object)
```

# Disscussion:
  The number of new COVID-19 infections peaked at the end of last year, and the number of new infections has been decreasing since the start of this year, when vaccination has been in full swing.
  In Denmark, the government started the lockdown at a relatively early stage and has been working to normalize the society as soon as possible by setting up a large scale inspection system.
  The "Corona-Pass" was introduced as a tool for this purpose.
  This is to indicate on the medical portal app that the patient is in one of the following conditions:
-Vaccinated
-Negative coronary test taken within 72 hours
-At least 14 days have passed since the positive test.

  Without this, person will not be allowed to enter restaurants, beauty salons, libraries, etc.
  
  In Japan, it is often expensive to issue a negative certificate, and there is hardly any way to prove to the outside world that you have been tested for COVID-19. I thought that such a system should be introduced in Japan.
