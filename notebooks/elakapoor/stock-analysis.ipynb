# Moving Average

```python
df_se['MA50'] = df_se.Open.rolling(50).mean()
df_se['MA100'] = df_se.Open.rolling(100).mean()
df_se['MA200'] = df_se.Open.rolling(200).mean()
```

```python
fig = go.Figure(data=[go.Scatter(x=df_se[df_se["Region"] == "Europe"]["Date"],
                                 y=df_se[df_se["Region"] == "Europe"]["Open"], 
                                 line = dict(color = "blue"), name="Open"), 
                      go.Scatter(x=df_se[df_se["Region"] == "Europe"]["Date"],
                                 y=df_se[df_se["Region"] == "Europe"]["MA50"], 
                                 line=dict(color='orange', width=2), name="MA50"),
                      go.Scatter(x=df_se[df_se["Region"] == "Europe"]["Date"],
                                 y=df_se[df_se["Region"] == "Europe"]["MA100"], 
                                 line=dict(color='green', width=2), name="MA100"),
                     go.Scatter(x=df_se[df_se["Region"] == "Europe"]["Date"],
                                 y=df_se[df_se["Region"] == "Europe"]["MA200"], 
                                 line=dict(color='magenta', width=2), name="MA200")])

fig.update_layout(
    plot_bgcolor = "white",
    title_text="Open Price and its moving average for Europe Exchange"
)

fig.update_yaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black",
                title_text="<b>Open and Moving Average</b>")
fig.update_xaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black", 
                 title_text="<b>Date</b>")



iplot(fig)
```

```python
fig = go.Figure(data=[go.Scatter(x=df_se[df_se["Region"] == "Switzerland"]["Date"],
                                 y=df_se[df_se["Region"] == "Switzerland"]["Open"], 
                                 line = dict(color = "blue"), name="Open"), 
                      go.Scatter(x=df_se[df_se["Region"] == "Switzerland"]["Date"],
                                 y=df_se[df_se["Region"] == "Switzerland"]["MA50"], 
                                 line=dict(color='orange', width=2), name="MA50"),
                      go.Scatter(x=df_se[df_se["Region"] == "Switzerland"]["Date"],
                                 y=df_se[df_se["Region"] == "Switzerland"]["MA100"], 
                                 line=dict(color='green', width=2), name="MA100"),
                     go.Scatter(x=df_se[df_se["Region"] == "Switzerland"]["Date"],
                                 y=df_se[df_se["Region"] == "Switzerland"]["MA200"], 
                                 line=dict(color='magenta', width=2), name="MA200")])

fig.update_layout(
    plot_bgcolor = "white",
    title_text="Open Price and its moving average for Swiss Exchange"
)

fig.update_yaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black",
                title_text="<b>Open Price and its Moving Average</b>")
fig.update_xaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black", 
                 title_text="<b>Date</b>")



iplot(fig)
```

Moving average is done to bring out smoothness in the curve. Howevere if moving average is more than it will miss the dips as seen in the case of MA200. So while calculating moving average it should be kept in mind that the value must not be too large.

# Stock Volatilitiy

```python
df_se["Return"] = (df_se["Close"]/df_se["Close"].shift(1)) - 1
```

```python
fig = px.histogram(df_se, x = "Return", color = "Region",color_discrete_map={
                "Europe":"#4363d8",
                "Switzerland":"#e6194B",
             })

fig.update_layout(
    plot_bgcolor = "white",
    title_text="Stock Volatility of Swiss and EuropeExchange"
)

fig.update_yaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black",
                title_text="<b>Value</b>")
fig.update_xaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black", 
                 title_text="<b>Return</b>")
iplot(fig)
```

```python
x2 = df_se[df_se["Region"] == "Europe"]["Return"]

x2_new = []
for x in x2:
    if not pd.isna(x):
        x2_new.append(x)
x2_new = pd.Series(x2_new)        
```

```python
x1 = df_se[df_se["Region"] == "Switzerland"]["Return"] 

x2 = df_se[df_se["Region"] == "Europe"]["Return"]
x2 = x2_new

hist_data = [x1, x2]

group_labels = ['Switzerland', 'Europe']
colors = ["#e6194B", '#4363d8']

fig = ff.create_distplot(hist_data, group_labels, show_hist=False, colors=colors)

fig.update_layout(
    plot_bgcolor = "white",
    title_text="Stock Volatility of Swiss and EuropeExchange"
)

fig.update_yaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black")
fig.update_xaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black", 
                 title_text="<b>Return</b>")

iplot(fig)
```

The density of both the exchange curve is nearly the same showing that they have nearly same stability.

# Cumulative Return

```python
df_se["Cumulative Return"] = (1 + df_se["Return"]).cumprod()
```

```python
fig = px.line(df_se, x = "Date", y = "Cumulative Return", color = "Region",color_discrete_map={
                "Europe":"#4363d8",
                "Switzerland":"#e6194B",
             })
fig.update_layout(
    plot_bgcolor = "white",
    title_text="Cumulative Return in Swiss and Europe Exchange"
)

fig.update_yaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black",
                title_text="<b>Cumulative Return</b>")
fig.update_xaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black", 
                title_text = "<b>Date</b>")
iplot(fig)
```

# Coorelation between both exchange

```python
df_cum = df_se[["Date", "Region", "Cumulative Return"]]
```

```python
df_cum_m = df_cum.pivot(index = "Date", columns='Region', values='Cumulative Return')
```

```python
fig = px.density_heatmap(df_cum_m, x="Switzerland", y="Europe", marginal_x="histogram", marginal_y="histogram")
fig.update_layout(
    plot_bgcolor = "white",
    title_text="Cumulative Return in Swiss and Europe Exchange"
)
iplot(fig)
```

So now we can see that if two stocks are positivley correlated with each other a linear relationship bewteen its cumulative return values should occur.

```python
import plotly.figure_factory as ff
fig = ff.create_scatterplotmatrix(df_cum_m, diag='histogram', 
                                  height=800, width=800)

fig.update_layout(
    plot_bgcolor = "white",
    title_text="Cumulative Return in Swiss and Europe Exchange"
)

iplot(fig)
```
