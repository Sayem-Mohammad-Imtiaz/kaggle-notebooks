# Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Importing Dataset 

```python
df_covid = pd.read_csv("/kaggle/input/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv")
df_covid.head()
```

#### Since, no. of rows are equal to no. of unique elements there are no duplicates in 'State/UTs' column

# Top 10 States with highest total cases

```python
high_cases = df_covid.sort_values(by='Total Cases',ascending=False).head(10)
xpos = np.arange(len(high_cases["State/UTs"]))
fig = plt.subplots(figsize=(15,6))
plt.xticks(xpos,high_cases["State/UTs"])
plt.bar(high_cases["State/UTs"],high_cases["Total Cases"],color = 'red',width=0.5)
plt.xlabel("Name of the State")
plt.ylabel("Total Cases")
plt.title("Top 10 states by Total cases")
plt.show()
```

# Top 5 States with least total cases

```python
least_cases = df_covid.sort_values(by='Total Cases',ascending=True).head()
ypos = np.arange(len(least_cases["State/UTs"]))
fig = plt.subplots(figsize=(20,6))
plt.xlabel("Name of the state")
plt.ylabel("Total Cases")
plt.title("Top 5 states having least cases")
plt.xticks(ypos,least_cases["State/UTs"])
plt.bar(least_cases["State/UTs"],least_cases["Total Cases"],color = 'orange',width=0.5)
plt.show()
```

# Top 10 states having highest number of deaths

```python
high_deaths = df_covid.sort_values(by = "Deaths",ascending = False).head(10)
state_names = np.arange(len(high_deaths["State/UTs"]))
plt.figure(figsize=(15,8))
plt.scatter(x = high_deaths["State/UTs"],y=high_deaths["Deaths"],c='black')
plt.plot(high_deaths["State/UTs"],high_deaths["Deaths"],c='r')
plt.xlabel("Name of State")
plt.ylabel("Deaths")
plt.title("Top 10 states with highest No. of Deaths")
plt.grid()
plt.show()
```

# Top 10 states having highest death ratio

```python
high_death_ratio = df_covid.sort_values(by='Death Ratio (%)',ascending=False).head(10)
states = np.arange(len(high_death_ratio["State/UTs"]))
sns.relplot(high_death_ratio["Total Cases"],high_death_ratio["Death Ratio (%)"],
            hue=high_death_ratio["State/UTs"])
plt.show()
```
