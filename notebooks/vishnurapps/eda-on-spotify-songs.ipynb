# Univariate Analysis

## Probability density function

```python
sns.FacetGrid(songs_data,height = 10)\
 .map(sns.distplot,'Beats.Per.Minute')\
 . add_legend();
plt.title("PDF of BPM")
plt.ylabel("Probablity")
plt.grid()
plt.show()
```

- Popular songs have a beat between 80 and 105

```python
sns.FacetGrid(songs_data,height = 10)\
 .map(sns.distplot,'Energy', bins=np.linspace(0, 100, 50))\
 . add_legend();
plt.title("PDF of energy")
plt.ylabel("Probablity")
plt.grid()
plt.show()
```

- Highly popular songs has an energy between 60 and 68

```python
sns.FacetGrid(songs_data,height = 10)\
 .map(sns.distplot,'Danceability', bins=np.linspace(0, 100, 50))\
 . add_legend();
plt.title("PDF of Danceability")
plt.ylabel("Probablity")
plt.grid()
plt.show()
```
