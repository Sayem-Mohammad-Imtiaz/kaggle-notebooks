This assignment is divided into two parts.

In the first part of the assignment, a global animation of COVID vaccination progress per country is mapped. 
The column 'people_fully_vaccinated_per_hundred' is selected to visualize the ratio between population fully immunized and total population up to the date in a country.
Fully immunized means someone who have received all doses prescribed by the vaccination protocol.



```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df = pd.read_csv("../input/covid-world-vaccination-progress/country_vaccinations.csv", header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
# print(df.columns)
fig = px.choropleth(df, locations='country', locationmode='country names', color='people_fully_vaccinated_per_hundred', hover_name='country', animation_frame='date', color_continuous_scale='rainbow', range_color=(0.0, 40.0))
fig.update_layout(title_text="Percentage of the country's total population that are fully vaccinated", title_x=0.45)
fig.show()

plotly.offline.plot(fig, filename='vaccinations.html')
```

# Global map of earthquake occurences

In the second part of the assignment, an hourly animation of 4.5+ magnitude earthquake occurrences from the end of July 2020 to October 2020 at Alaska is mapped. Alaska has been chosen as the target region because the July 22nd Alaska Peninsula earthquake was the strongest earthquake of 2020 by magnitude. It recorded a magnitude of 7.8. A huge aftershock of this earthquake happened almost 3 months later on 19th October. This 7.6-magnitude aftershock became the third strongest earthquake of 2020. Fortunately, there were no casualties nor fatalities from both of the events.



**Source:**

> Grassi, B., Gardine, L., & Ruppert, N. (2021, February 10). *2020 Seismicity Year in Review*. Alaska Earthquake Center. https://earthquake.alaska.edu/2020-seismicity-year-review

```python
df = pd.read_csv('../input/2020-alaska-earthquake/query.csv')
df.index = pd.to_datetime(df['time'])
df['time'] = df.index.strftime('%Y-%m-%d %H:00:00')
df = df.sort_index()
fig = px.scatter_geo(df, lat='latitude', lon='longitude', color='mag', animation_frame='time', color_continuous_scale='rainbow', range_color=(4.5, 8.0))
# southern alaska coordinates
lat_foc = 55.704037 
lon_foc = -159.599384
fig.update_layout(
        geo = dict(
            projection_scale=8,
            center=dict(lat=lat_foc, lon=lon_foc)
        ))

# fig.update_layout(geo = dict(scope='north america'))

fig.show()
```
