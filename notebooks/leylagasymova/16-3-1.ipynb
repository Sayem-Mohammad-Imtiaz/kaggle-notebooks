```python
import pandas as pd
import statistics as st
import matplotlib.pyplot as plt
import numpy as np

#1.Найти готовый файл с данными 
VG=pd.read_csv('../input/videogamesales/vgsales.csv')
VG
```

```python
#2.Посчитать статистические показатели:
#- медиана
print("Медиана= ", np.median(VG.Global_Sales))

#- мода
print("Мода= ", st.mode(VG.Global_Sales))

#- средняя
print("Среднее= ", np.std(VG.Global_Sales))

#- минимум
print("Минимум= ", np.min(VG.Global_Sales))

#- максимум
print("Максимум= ", np.max(VG.Global_Sales))
```

```python
#3.Создать новый столбец, в котором разместить разницу между максимумом и минимумом.
VG['Difference']=VG.max(axis = 1) - VG.min(axis = 1)
VG
```

```python
#4.Разбить данные на несколько блоков (не менее 3, приблизительно равных по размеру) по некоторому критерию.
#разбиваю по годам выпуска игр 
oldGame=0
newerGame=0
newGame=0

for index, row in VG.iterrows(): 
    if row[3]<=2003:
        oldGame=oldGame+1
    if row[3]>2003 and row[3]<=2009:
        newerGame=newerGame+1
    if row[3]>2009:
        newGame=newGame+1

col=VG['Name'].count()
a1=oldGame/col
a2=newerGame/col
a3=1-a1-a2


#5.Привести круговую диаграмму для проделанного разбиения.
labels = ['older than 2003','from 2003 to 2009','from 2009']
values = [a1,a2,a3]
colors = ['yellow','green','red']
plt.title('release of games')
plt.pie(values,labels=labels,colors=colors, shadow=True,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
#6.В зависимости от особенностей набора данных для каждого блока построить график (гистограмму).
oldVG=VG[(VG['Year']<=2003)]
newerVG=VG[(VG['Year']>2003)&(VG['Year']<=2009)]
newVG=VG[(VG['Year']>2009)]

plt.hist(oldVG['Year'])
plt.show() 
plt.scatter(oldVG['Year'], oldVG['Global_Sales'])
plt.show()

plt.hist(newerVG['Year'])
plt.show()
plt.scatter(newerVG['Year'], newerVG['Global_Sales'])
plt.show()

plt.hist(newVG['Year'])
plt.show()
plt.scatter(newVG['Year'], newVG['Global_Sales'])
plt.show()
```
