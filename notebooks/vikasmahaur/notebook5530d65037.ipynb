```python
import matplotlib.pyplot as plt
import numpy as np
import os
import pandas as pd
```

```python
df = pd.read_csv("../input/suicide-rates-overview-1985-to-2016/master.csv")
```

```python
df.head()
```

```python
df=df.dropna(axis="columns") # cleaning data, remove columns with nan
df.head()
```

```python
df_sg=df.loc[df["country"]=="Singapore"]
df_sg.head()
```

**nvestigate general suicide trend, check year to total suicide/100k
pop and gdp. Attempt to combine suicide no for all ages, male and
female.
**

```python
simplified=df_sg.drop(["country","sex","age","country-year","generation"], axis= "columns")
simplified
```

**Attempt to group by year and sum the values together**

```python
sg_year=simplified.groupby("year")[["suicides_no","population"]].sum() # year, suicideno,population
sg_year.head()
```

```python
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
simplified.columns
simplified[" gdp_for_year ($) "].dtype
```

**Gdp for year columns is an object, not a number. thats why its not
calulated in the mean. To change it to a number i need to remove
to commas, and change the type to an integer**

```python
simplified[" gdp_for_year ($) "]=simplified[" gdp_for_year ($) "].str.replace(',',"")
sg_capita=simplified.groupby("year")[[" gdp_for_year ($) ","gdp_per_capita ($)"]].mean()
sg_capita.head()
```

```python
sg_finaldf=sg_year.join(sg_capita)
sg_finaldf.head()
```

```python
sg_finaldf["suicide/pop"]=sg_finaldf["suicides_no"] / sg_finaldf["population"]
sg_finaldf.head()
```

```python
sns.lineplot(data=sg_finaldf,x=sg_finaldf["gdp_per_capita ($)"],y=sg_finaldf["suicide/pop"])
sns.regplot(x=sg_finaldf["gdp_per_capita ($)"],y=sg_finaldf["suicide/pop"])
```
