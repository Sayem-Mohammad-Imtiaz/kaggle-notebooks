# label encoding all the categorical features

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
for col in data.columns:
    data[col] = labelencoder.fit_transform(data[col])
 
```

```python
data.head()
```

## Giving the correlation plot to get the idea of correlation between features

```python


corrmat = data.corr()
plt.subplots(figsize=(12,9))
sns.heatmap(corrmat,  square=True,cbar=True)

```

so what we get:
1. veil-color has no effect on anyone i.e it has to be dropped.
2. gill size is mostly correlated with class

```python
data=data.drop('veil-type',axis=1)
y=data['class'].values
data=data.drop('class',axis=1)
```

```python
y
```

lets perform splitting the dataset

```python


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data,y,test_size=0.2,random_state=4)

```

applying the most efficient classifier: xgboost model

```python

from xgboost import XGBClassifier
classifier=XGBClassifier()

```

```python

classifier.fit(X_train, y_train)
```

## you can tune the parameters as per your choice

```python
classifier.score(X_train, y_train)
```

```python

from sklearn.model_selection import cross_val_score

scores = cross_val_score(classifier, X_train, y_train, cv=10, scoring='accuracy')

```

```python
scores
scores.mean()
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn import metrics
```

```python
auc_roc=metrics.roc_auc_score(y_test,y_pred)
auc_roc
```

```python

from sklearn import metrics

confusion_matrix=metrics.confusion_matrix(y_test,y_pred)
confusion_matrix
```
