```python
df = pd.read_csv("../input/pitching.csv")
df_sg = df[(df.gs == df.g)]
```

```python
Y = df_sg.w/(df_sg.gs)
Y_class = np.floor(Y)
```

```python
X = df.drop(['player_id','year','stint','team_id','league_id'],axis=1)
pl = ~np.isnan(X).any(axis=1)
X = X[pl]
Y_class = Y_class[]
```

```python
from sklearn import svm
```

```python
clf = svm.SVC()
clf.fit(X, Y_class) 
```
