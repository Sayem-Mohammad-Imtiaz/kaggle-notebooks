```python
labelencoder = LabelEncoder()
iris_data["species"] = labelencoder.fit_transform(iris_data["species"])
iris_data

#iris_data['species'] = iris_data['species'].map({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
```

```python
X = iris_data.drop("species", axis=1)
Y = iris_data['species']
```

```python
X.head()
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.30 , random_state = 2)
```

```python
#X_train
```

```python
#Y_train
```

```python
#X_test
```

```python
for k in range(1,20):
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train,Y_train)
    prediction = model.predict(X_test)
    accuracy = accuracy_score(Y_test,prediction)
    print("For K =",k, "the accuracy is:",accuracy)
```

```python
Best_model = KNeighborsClassifier(n_neighbors=1)
Best_model.fit(X_train,Y_train)
prediction = Best_model.predict(X_test)
accuracy = accuracy_score(Y_test,prediction)
print("For k = ",1," Validation accuracy is:",accuracy)

```

```python
Train_prediction = Best_model.predict(X_train)
Train_prediction
# training accuracy
accuracy_score(Y_train,Train_prediction)
```
