### 18. Global vs Local

Global variables are the one that are defined and declared outside a function and we need to use them inside a function. We only need to use the global keyword in a function if we want to do assignments / change them. global is not needed for printing and accessing.

John Wick knew this. That's why he was able to beat the Boogeyman. Boogeyman didn't know it.

![](https://i.imgur.com/Cyyxwc4.png)

```python
#Error because BestHitman is being assigned a value but was never defined as global

BestHitman = 'Boggeyman'

def whoIsBestHitman():
    
    print("Best hitman was ",BestHitman)
    BestHitman = 'John Wick' 
    print("But now it is ",BestHitman)
    
whoIsBestHitman()
```

```python
BestHitman = 'Boggeyman'

def whoIsBestHitman():
    
    global BestHitman
    
    print("Best hitman was ",BestHitman)
    BestHitman = 'John Wick' 
    print("But now it is ",BestHitman)
    
whoIsBestHitman()
```

### 36. DeafultDict() from collections
The functionality of both dictionaries and defualtdict are almost same except for the fact that defualtdict never raises a KeyError. It provides a default value for the key that does not exists.

```python
from collections import defaultdict 
  
# Function to return a default 
# values for keys that is not 
# present 
def def_value(): 
    return "Not Present"
      
# Defining the dict 
d = defaultdict(def_value) 
d["a"] = 1
d["b"] = 2
  
print(d["a"]) 
print(d["b"]) 
print(d["c"]) 
```

### 37. Ordered Dictionary from collections
OrderedDict preserves the order in which the keys are inserted. A regular dict doesnâ€™t track the insertion order, and iterating it gives the values in an arbitrary order and won't give ordered representation everytime. This is noticeable when the dict size is large.

![](https://i.imgur.com/YXdHQAk.png)

```python
# A Python program to demonstrate working of OrderedDict 
from collections import OrderedDict 
    
print("This is an Ordered Dict:\n") 
od = OrderedDict() 

od = {i : i for i in range(10)}

for key, value in od.items(): 
    print(key, value) 
```

### 47. Check worth of info

This is rather a very interesting topic I discovered when I was reading materials online to answer a [discussion](https://www.kaggle.com/learn-forum/148901#835928) question.
As the name suggests, this is a hack to find a quantitative value of how much of information is generated by 'n' columns that are highly corelated.

```python
corr_list = list(so_schema.loc[col_to_choose]['QuestionText'])

print("The whole text is \n" + corr_list[0])
print()
print("Trimmed text \n" + corr_list[0][240:])
```

```python
df_combined[col_to_choose].corr()
```

```python
df_combined[['AssessJob1','AssessJob9']].corr()
```

```python
cols_selected_squared = 4 #Because we are choosing two columns, 2^2 = 4

info_worth = cols_selected_squared/(df_combined[['AssessJob1','AssessJob9']].corr()**2).sum().sum()
print("The two columns have given you {0:.2f} columns worth of info".format(info_worth))
```

```python
info_worth = cols_selected_squared/(df_combined[['AssessJob1','AssessJob1']].corr()**2).sum().sum()
print("The two columns have given you {0:.2f} columns worth of info".format(info_worth))
```
