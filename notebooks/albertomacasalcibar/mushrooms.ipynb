```python
class_rep = {'p':'Poisonous', 'e': 'Edible'}
cap_shape_rep = {'x':'Convex', 'b':'Bell', 's':'Sunken', 'f':'Flat', 'k':'Knobbed','c':'Conical'}
habitat_rep = {'u': 'Urban','g': 'Grasses','m': 'Meadows','d': 'Woods','p': 'Paths','w': 'Waste','l': 'Leaves'}
```

```python

df['habitat'] = df['habitat'].replace(habitat_rep)
columns = df.columns.tolist()
```

```python
for column in columns:
    if df[column].nunique() >= 6:
        sns.countplot(df[column], order = df[column].value_counts().index)
    else:
        plt.pie(df[column].value_counts(), labels = df[column].unique())
    plt.title(column.title() + ' Value Counts')
    plt.xticks(rotation=30, fontsize=10)
    plt.xlabel(column.title(), fontsize=12) 
    plt.show()
    plt.clf()
```
