## Where are vaccinated more people per day? But in terms of percent from entire population ?

```python
df.columns
```

```python
df1=pd.crosstab(df['date'],df['country'],values=df['daily_vaccinations_per_million'],aggfunc=sum)
```

```python
for i in df1.columns:
    df1[i].plot(kind='line',figsize=(12,8),marker='o')
    plt.xlabel(" Date ")
    plt.ylabel("Number of People Vaccinated")
    plt.title(i,fontsize=18)
    plt.savefig(i,dpi=300,facecolor=None)
    plt.show()
```
