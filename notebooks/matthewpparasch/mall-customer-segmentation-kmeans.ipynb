```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
```

```python
plt.style.use('seaborn')
```

```python
df = pd.read_csv(r'/kaggle/input/customer-segmentation-tutorial-in-python/Mall_Customers.csv')
```

## Initial review of data
Checking for nulls and distribution of the features

```python
df.info()
```

```python
df.describe()
```

```python
features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
fig, axs = plt.subplots(len(features),1, figsize=(10,10))

for i, v in enumerate(features):
    sns.distplot(df[v], ax=axs[i], bins=20)
    axs[i].set_title('{} Distribution'.format(v))

plt.tight_layout()
plt.show()
```

## Multivariable Analysis
Looking for trends between spending score and other features

```python
features.remove('Spending Score (1-100)')
```

```python
fig, axs = plt.subplots(len(features),1, figsize=(8,8))

for i,v in enumerate(features):
    sns.scatterplot(df['Spending Score (1-100)'], df[v], ax=axs[i])
    axs[i].set_title(v)

plt.tight_layout()
plt.show()
```

```python
sns.jointplot(df['Annual Income (k$)'], df['Age'])
```

## Creating segments based on variable relationships
After reviewing the data, it looks like there are 5 clear groupings in the relationship between income and spending score. There are what looks like 4 (less defined) groups in the age vs spending score relationship

```python
X = df.loc[:,['Spending Score (1-100)', 'Annual Income (k$)']].values

km = KMeans(n_clusters=5)
km.fit(X)
c = km.cluster_centers_
```

```python
color = np.array(['red', 'blue', 'green', 'yellow', 'purple'])
labels = km.labels_

fig, axs = plt.subplots(figsize=(8,8))
fig.suptitle('KMeans Clustering', fontsize=22)

axs.scatter(x=X[:,0], y=X[:,1], c=color[km.labels_], s=200)
axs.scatter(x=c[:,0], y=c[:,1], marker='X', s=500, c='black')

axs.set_xlabel('Spending Score (0-100)', fontsize=16)
axs.set_ylabel('Income (k$)', fontsize=16)

plt.tight_layout(rect=[0,0,1,.95])
plt.show()
```

```python
X = df.loc[:,['Spending Score (1-100)', 'Age']].values
km = KMeans(n_clusters=4)
km.fit(X)
c = km.cluster_centers_
```

```python
labels = km.labels_

fig, axs = plt.subplots(figsize=(8,8))
fig.suptitle('KMeans Clustering', fontsize=22)

axs.scatter(x=X[:,0], y=X[:,1], c=color[km.labels_], s=200)
axs.scatter(x=c[:,0], y=c[:,1], marker='X', s=500, c='black')

axs.set_xlabel('Spending Score (0-100)', fontsize=16)
axs.set_ylabel('Age', fontsize=16)

plt.tight_layout(rect=[0,0,1,.95])
plt.show()
```
