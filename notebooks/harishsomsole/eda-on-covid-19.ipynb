# Step 1 ::  Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

# Step 2 :: Load Dataset

```python
df = pd.read_csv("/kaggle/input/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv")
```

# Total Cases statewise

```python
plt.figure(figsize=(16,6))
px.bar(df,x='State/UTs',y='Total Cases',title='Total cases in All States/UTs (India)',log_y=True)
```

Maharastra State has more covid cases 

Andaman and Nicobar has less cases  

# Total Deaths Statewise

```python
plt.figure(figsize=(16,6))
px.bar(df,x='State/UTs',y='Deaths',title='Total Death cases in All States/UTs (India)',log_y=True)
```

Maharastra State has high deaths due to more cases
and Daman and Diu has less deaths than Andaman and Nicobar

# Recovered cases Statewise

```python
plt.figure(figsize=(16,6))
px.bar(df,x='State/UTs',y='Discharged',title='Total Recoverd/Discharged cases in All States/UTs (India)',log_y=True)
```

# Active cases statewise

```python
plt.figure(figsize=(16,6))
px.bar(df,x='State/UTs',y='Active',title='Total Active cases in All States/UTs (India)',log_y=True)
```

# % ratio's of discharge,active and deaths statewise

```python
plt.figure(figsize=(16,6))
px.bar(df,x='State/UTs',y='Discharge Ratio (%)',title='Discharge ratio in % All States/UTs (India)',log_y=True)
```

Mizoram state has less Discharge percentage 

Daman and Diu has high discharge percentage

```python
plt.figure(figsize=(16,6))
px.bar(df,x='State/UTs',y='Death Ratio (%)',title='Discharge ratio in % All States/UTs (India)',log_y=True)
```

```python
plt.figure(figsize=(16,6))
px.bar(df,x='State/UTs',y='Active Ratio (%)',title='Active ratio in % All States/UTs (India)',log_y=True)
```

Punjab has highest death percentage and also its clear that no state has greater than 3% deaths of total cases

# Relation Between Features

```python
cor = df.corr()
plt.figure(figsize=(16,6))
sns.heatmap(cor,annot=True)
```

# State wise Active, Discharged , Total, Deaths cases

```python
# xl = ['Total Cases','Active','Discharged']

for su in df['State/UTs'].values:
    data = df[['Total Cases','Active','Discharged','Deaths']][df['State/UTs'] == su]
    data.plot(kind='bar',log=True,title=su)
```
