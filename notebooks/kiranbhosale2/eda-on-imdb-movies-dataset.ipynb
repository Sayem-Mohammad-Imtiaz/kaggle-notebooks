```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv")
```

```python
df.head(5)
```

```python
df.info()
```

# Number of null values

```python
df.isnull().sum()
```

```python
df=df.rename(columns={'Series_Title':'Movie_Title'})
```

```python
df.head(3)
```

```python
df["Gross"]
```

## Gross data is in object type need to convert in int.

```python
df['Gross']=df['Gross'].str.replace(',','')
df['Gross'] = df['Gross'].replace(np.nan, 0)
df['Gross'] = df['Gross'].astype(int)
```

```python
df["Gross"].dropna(inplace=True)
df["Gross"].head()
```

```python
top_gross = df.sort_values(['Gross'], ascending = False)
top_gross.head(3)
```

```python
fig,axs=plt.subplots(figsize=(20,5))
bar=sns.barplot(x=top_gross['Movie_Title'][:6],y=(top_gross['Gross']/1000000)[:6],palette = 'hls')
bar.set(xlabel='Movie Title', ylabel='Gross in Millions')
bar.set_title("Top 6 Movies by Gross",weight='bold')
```

# Playing with top movies (by voting).

```python
top_movies=df.sort_values('No_of_Votes', ascending=False)
```

```python
fig,axs=plt.subplots(figsize=(15,5))
rat=sns.barplot(x=top_movies['Movie_Title'][:6],y=(top_movies['No_of_Votes']/1000000)[:6],palette = 'husl')
rat.set(xlabel='Movie Title', ylabel='No of votes in millions')
rat.set_title("TOP 6 most Voted movies", weight = "bold")
plt.show()
```

### Ratings of Top voted movies. 

```python
fig,axs=plt.subplots(figsize=(15,5))
rat=sns.barplot(x=top_movies['Movie_Title'][:6],y=top_movies['IMDB_Rating'][:6],palette = 'Set2')
rat.set_title("TOP 6 IMDB Rating movies", weight = "bold")
plt.show()
```

# Observations with run time.
need to convert runtime in int format.

```python
runtime=df['Runtime']
runtime.shape
```

```python
runtime=runtime.str.replace(' min','')
runtime=runtime.astype('int')
```

```python
fig,axs=plt.subplots(figsize=(20,5))
d=sns.kdeplot(runtime)
d.set_title("Time Duration of movies", weight = "bold")
```
