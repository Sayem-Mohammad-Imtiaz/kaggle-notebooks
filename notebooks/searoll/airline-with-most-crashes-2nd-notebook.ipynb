**As a flight student, I've always been faschinated with which airlines and region is the most dangerous ones. In this notebook, I will try to find out which airline has the most accidents**

```python
# importing useful libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing data
crash_data = pd.read_csv('../input/airplane-crash-data-since-1908/Airplane_Crashes_and_Fatalities_Since_1908_20190820105639.csv')

# convert dates to datetime
crash_data['Date'] = pd.to_datetime(crash_data['Date'])

# As we will not need operators which are Private, Mail services or Military, we will remove those
crash_data = crash_data[crash_data['Operator'].str.contains("Military|Private|Mail") == False]

crash_data.describe()
```

```python
# Finding null values
sns.heatmap(crash_data.isnull(), cbar=False)
```

**There seems to be a lot of missing values (especially in Flight Number). The case with flight number could be because of most operators being military or private. As most values won't be really relevant, We will leave them out**

```python
# get relevant data
cleaned_operator_data = crash_data[crash_data['Operator'].notna()]
operator_by_crash = cleaned_operator_data.groupby('Operator').size().sort_values(ascending=False)

# plot bar charts
plt.figure(figsize=(24, 7))
sns.barplot(x=operator_by_crash.head(7).index, y=operator_by_crash.head(7))
plt.title('Most crashes per airline')
plt.ylabel('Nr of Crashes')
plt.xlabel('Operator')
```

**Out from this chart, Aeroflot is the airline with most crashes by a margin.**

```python
# Getting the most recent airline crashes from 2000
crash_data_from_2000 = cleaned_operator_data[pd.DatetimeIndex(cleaned_operator_data['Date']).year >= 2000]
operator_by_crash = crash_data_from_2000.groupby('Operator').size().sort_values(ascending=False)

# Plot barchart 
plt.figure(figsize=(24, 7))
sns.barplot(x=operator_by_crash.head(10).index, y=operator_by_crash.head(10))
plt.title('Most crashes per airline from year 2000')
plt.ylabel('Nr of Crashes')
plt.xlabel('Operator')
```

**From this data, we can conclude that Aeroflot has improved a lot in the last year  with not being top 10 anymore**
