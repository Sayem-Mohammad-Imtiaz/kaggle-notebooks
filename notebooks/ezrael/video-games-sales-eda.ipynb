## Importing Library

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

## Loading Dataset

```python
PATH = "/kaggle/input/videogamesales/vgsales.csv"
```

```python
df = pd.read_csv(PATH)
```

## Analyzing

```python
import matplotlib.pyplot as plt
```

### Global Sales by Platform

```python
### Global sales by platform

globalByPlatform = dict(df.groupby("Platform")["Global_Sales"].mean())
k, v = zip(*globalByPlatform.items())

plt.figure(figsize=(16, 7))
plt.bar(k, v);
```

Game Boy had the most sales average in global level, beside that NES also had high sales in global level 

### Country Region Sales by Platform

```python
### Sales by platform

NAPltfm = df.groupby('Platform')['NA_Sales'].mean()
EUPltfm = df.groupby('Platform')['EU_Sales'].mean()
JPPltfm = df.groupby('Platform')['JP_Sales'].mean()
OtherPltfm = df.groupby('Platform')['Other_Sales'].mean()

labels, NAv = zip(*NAPltfm.items())
_, EUv = zip(*EUPltfm.items())
_, JPv = zip(*JPPltfm.items())
_, Otherv = zip(*OtherPltfm.items())


x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize=(20, 8))
rects1 = ax.bar(x - width, NAv, width, label='NA')
rects2 = ax.bar(x - width/2, EUv, width, label='EU')
rects3 = ax.bar(x + width/2, JPv, width, label='JP')
rects4 = ax.bar(x + width, Otherv, width, label='Other')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Sales mean')
ax.set_title('Sales by platform')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend();
```

NES had the most average sales in North America Region and Japan Region, meanwhile in Europe Region, GameBoy had the most average sales, in other regionn PS4 had the most average sales.

### Global sales by genre

```python
globalByGenre = dict(df.groupby("Genre")["Global_Sales"].mean())
k, v = zip(*globalByGenre.items())

plt.figure(figsize=(16, 7))
plt.bar(k, v);
```

Video Games with Platform genre had the most sales globaly

```python
### Sales by genre

NAPltfm = df.groupby('Genre')['NA_Sales'].mean()
EUPltfm = df.groupby('Genre')['EU_Sales'].mean()
JPPltfm = df.groupby('Genre')['JP_Sales'].mean()
OtherPltfm = df.groupby('Genre')['Other_Sales'].mean()

labels, NAv = zip(*NAPltfm.items())
_, EUv = zip(*EUPltfm.items())
_, JPv = zip(*JPPltfm.items())
_, Otherv = zip(*OtherPltfm.items())


x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots(figsize=(20, 8))
rects1 = ax.bar(x - width, NAv, width, label='NA')
rects2 = ax.bar(x - width/2, EUv, width, label='EU')
rects3 = ax.bar(x + width/2, JPv, width, label='JP')
rects4 = ax.bar(x + width, Otherv, width, label='Other')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Sales mean')
ax.set_title('Sales by genre')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend();
```

Platformer genre had the most sales in North America region, meanwhile in Japan and other region shooter genre games had the most sales, and finally in Europe region Role-Playing games had the most sales.
