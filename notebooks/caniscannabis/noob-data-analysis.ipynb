```python
df_america=df_dropped.loc[df_dropped['Production Country']=='United States']
```

```python
for i in df_america.index:
  df_america['Imdb Score'][i]=df_america['Imdb Score'][i].replace('/10','')
```

```python
df_america['Imdb Score']=df_america['Imdb Score'].astype(float)
```

```python
for i in df_america.index:
  if(df_america['Imdb Score'][i]>=5):
    df_america['Imdb Score'][i]=1
  else:
    df_america['Imdb Score'][i]=0  
```

```python
df_america['Imdb Score'].value_counts().plot(kind='pie',figsize=(15,15))
```
