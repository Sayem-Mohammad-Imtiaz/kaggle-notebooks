```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import os
```

```python
dataset = pd.read_csv("../input/salary-data-simple-linear-regression/Salary_Data.csv")
dataset = dataset.dropna()
dataset.head()
```

```python
dataset.info()
```

```python
x = np.array(dataset['YearsExperience']).reshape(-1,1)
x[:5]
```

```python
y = np.array(dataset['Salary'])
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.48)
```

```python
model = LinearRegression()
model.fit(x_train , y_train)
```

```python
y_pred = model.predict(x_test)
y[:5]
```

```python
plt.scatter(x_test , y_test , color = "green")
plt.plot(x_test,y_pred,color = "blue")
plt.title("Salary vs Experience")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
result = pd.DataFrame({'Years Experience': x_test[:,0], 'Actual Salary': y_test, 'Predicted Salary': y_pred})
result.head()
```

```python
print("Score : ")
model.score(x_test,y_test)
```
