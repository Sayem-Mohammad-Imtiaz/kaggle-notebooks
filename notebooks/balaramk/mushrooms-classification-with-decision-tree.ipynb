```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
def details(df):
    print('Head:-\n',df.head(3),'\n--------------------------------------------\n')
    print('Columns :-\n',df.columns,'\n--------------------------------------------\n')
    print('Desccription:-\n',df.describe,'\n--------------------------------------------\n')
```

#  Reading and Intro to the dataset 



```python
df = pd.read_csv(r'/kaggle/input/mushroom-classification/mushrooms.csv')
details(df)
```

# Categorical to numerical

As we see from the above ouput that the description of mushrooms are respective alphabets and obviously any ML algorithm does not work with alphabets and can understand only numbers. Hence we transform all the alphabets into numbers.

For this we use the label encoder from sklearn.

```python
labelencoder=LabelEncoder()
for column in df.columns:
    df[column] = labelencoder.fit_transform(df[column])
print("DataFrame after encoding :- \n\n")
print(df.head(2))
```

As we observe all the alphabetical descriptions have been changed into a corresponding number.


# Tiny Pre-Processing that matters a lot

**Checking for missing or null values**

```python
df.isna().sum()
```

```python
df.describe()
df['veil-type'][5]
# removing veil type bcoz all are 0
df=df.drop(["veil-type"],axis=1)
```

**We Removed the entire column of veil-type as all the values are 0**

# Splitting and Feature selection

```python
X = df.drop(['class'], axis=1) ; y = df[['class']]
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

# Building model with DecisionTree and Entropy as criterion

```python
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
acc = metrics.accuracy_score(y_pred,y_test)
print('Accuracy score = ',acc*100,'%')
```

**NOTE :- In this case there is no effect on the criterion on the accuracy score as both GINI and ENTROPY give out the same accuracy. You can also use criterion as GINI**

# CONFUSION MATRIX and CLASSIFICATION REPORT

```python
cf_matrix = metrics.confusion_matrix(y_test,y_pred)
_=sns.heatmap(cf_matrix, annot=True,fmt='d',cmap="Blues")
```

```python
print(metrics.classification_report(y_test,y_pred))
```
