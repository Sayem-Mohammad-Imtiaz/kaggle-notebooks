```python
# Here cut_points and label_names are used to categorise the age into different categories like infant,chind,adult etc
cut_points = [-1,0,5,12,18,35,60,100]
label_names = ["Missing","Infant","Child","Teenager","Young Adult","Adult","Senior"]
#filling Nan values with -0.5  represnts the missing case in age
train["Age"] = train["Age"].fillna(-0.5)
train["Age_categories"] =pd.cut(train["Age"],cut_points,labels=label_names)
train_pivot=train.pivot_table(index="Age_categories",values="Survived")
train_pivot.plot.bar()
plt.show()
```

Let's define a function which converts our required columns to numerical values

```python
def create_dummies(df,column_name):
    dummies = pd.get_dummies(df[column_name],prefix=column_name)
    df = pd.concat([df,dummies],axis=1)
    return df

train = create_dummies(train,"Pclass")
train = create_dummies(train,"Sex")
train = create_dummies(train,"Age_categories")
train.head()
```

```python
from sklearn.model_selection import train_test_split

columns = ['Pclass_1', 'Pclass_2', 'Pclass_3', 'Sex_female', 'Sex_male',
       'Age_categories_Missing','Age_categories_Infant',
       'Age_categories_Child', 'Age_categories_Teenager',
       'Age_categories_Young Adult', 'Age_categories_Adult',
       'Age_categories_Senior']
X_train,X_test,y_train,y_test=train_test_split(train[columns],train["Survived"],test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```
