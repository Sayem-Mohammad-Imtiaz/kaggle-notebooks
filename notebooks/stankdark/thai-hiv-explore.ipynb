```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

labels = [2019,2015,2010,2005,2000]
data_male   = df[df['Location']=="Thailand" ][df['Dim1']=="Male"]["First Tooltip"].tolist()
data_female = df[df['Location']=="Thailand" ][df['Dim1']=="Female"]["First Tooltip"].tolist()
data_both   = df[df['Location']=="Thailand" ][df['Dim1']=="Both sexes"]["First Tooltip"].tolist()

labels.reverse()
data_male.reverse()
data_female.reverse()
data_both.reverse()


for i in range(len(data_male)):
    data_male[i] = getValue(data_male[i])
    data_female[i] = getValue(data_female[i])
    data_both[i] = getValue(data_both[i])

x = np.arange(len(labels))  # the label locations
width = 0.3  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width, data_male, width, label='Male' , color="blue")
rects2 = ax.bar(x , data_female, width, label='Female' , color="red")
rects3 = ax.bar(x+ width, data_both, width, label='Both' , color='green')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('per 1000 uninfected pop.')
ax.set_yticks(np.arange(0.0, 1.0, 0.1))
ax.set_title('New HIV infections per 1000 uninfected population')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

fig.tight_layout()

plt.show()
```
