# Data visualisation 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class dataVis:
    '''This class to visualize data '''

    '''Histogram'''
    def hist(self): 
        global Data
        hist=Data.hist(figsize=(9,9))
        return hist
    
    '''density plot'''
    def density(self):
        global Data
        dens=Data.plot(kind='density', subplots=True, layout=(3,4), sharex=False, figsize=(9,9))
        return dens
    
    '''Pairplot'''
    def pairPlot(self):
        global Data
        pair=sns.pairplot(data=Data ,y_vars=["price"],hue="cut")
        return pair
    
    '''Boxplot'''
    def boxPlot(self):
        global Data
        box=Data.plot(kind='box', subplots=True, layout=(3,4), sharex=False, figsize=(9,9))
        return box
    
    '''Corrolation Matrix'''
    def corr(self): 
        global Data        
        plt.subplots(figsize=(10,8))
        correlations = Data.corr(method='pearson')
        corr=sns.heatmap(correlations, square=True, annot=True)
        return corr
    
    '''Scatter Matrix'''
    def scatter(self): 
        global Data        
        from pandas.plotting import scatter_matrix
        sc=scatter_matrix(Data, figsize=(9,9))
        return sc
    
    '''Pie Plot'''
    def piePlot(self): 
        global Data
        p=Data.groupby(['cut']).sum().plot(kind='pie', y='price',startangle=90,
        figsize=(8,8), autopct='%1.1f%%')
        return p
```

# Data Cleaning

```python
from sklearn.preprocessing import LabelEncoder

class DataCleaning:
    '''This class to preprocessing data '''
    

    def droppingZeros(self):
        global Data
        
        '''  2-Min value of "x", "y", "z" are zero this indicates
        that there are faulty values in data that represents dimensionless or 2-dimensional diamonds.
        So we need to filter it'''
        Data = Data.drop(Data[Data["x"]==0].index)
        Data = Data.drop(Data[Data["y"]==0].index)
        Data = Data.drop(Data[Data["z"]==0].index)
    
        return Data.shape
    
    def outliersData(self):
        '''This function to remove outliers values'''
        
        global Data
        Data = Data[(Data["depth"]<75)&(Data["depth"]>45)]
        Data = Data[(Data["table"]<80)&(Data["table"]>40)]
        Data = Data[(Data["x"]<30)]
        Data = Data[(Data["y"]<30)]
        Data = Data[(Data["z"]<30)&(Data["z"]>2)]
           
        return Data.shape
    
    def unNamedClos(self):
        '''This function to delete Unnamed colmuns which are indecis just '''
        global Data
        
        Data.drop(Data.columns[0],axis=1,inplace=True)
        return Data
    
    def Encoding(self):
        '''This function to convert categorial attaribute to numeric values'''
        
        global Data
        encoder=LabelEncoder()
        
        Data['color']=encoder.fit_transform(Data['color'])
        Data_mapping={index: label for index, label in enumerate(encoder.classes_)}
        
        Data['cut']=encoder.fit_transform(Data['cut'])
        Data_mapping={index: label for index, label in enumerate(encoder.classes_)}
        
        Data['clarity']=encoder.fit_transform(Data['clarity'])
        Data_mapping={index: label for index, label in enumerate(encoder.classes_)}
        
        return Data
```

```python
# Compare Algorithms
from pandas import read_csv
from matplotlib import pyplot
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

# prepare models
models = []
models.append(('LR', LinearRegression()))
models.append(('DT',DecisionTreeRegressor()))
models.append(('NB', GaussianNB()))
models.append(('KNN',KNeighborsClassifier()))
models.append(('RFR',RandomForestRegressor()))

#evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = KFold(n_splits=10)
    cv_results = cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
    results.append(cv_results)

    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
    
# boxplot algorithm comparison
fig = pyplot.figure(figsize=(12,12))
fig.suptitle('Algorithm Comparison')
ax = fig.add_subplot(111)
pyplot.boxplot(results)
ax.set_xticklabels(names)
pyplot.show()
```
