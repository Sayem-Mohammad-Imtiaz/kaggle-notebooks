```python
X = heart_data.drop(['output'], axis=1)
y = heart_data.output

categorical_features = ['sex', 'cp','fbs','restecg', 'exng', 'slp', 'caa', 'thall']
X = pd.get_dummies(X, columns=categorical_features)
```

```python
from sklearn.model_selection import train_test_split, learning_curve, cross_val_score


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)


from sklearn.preprocessing import StandardScaler

continuous_features = ['age', 'oldpeak', 'trtbps','chol', 'thalachh']
scaler = StandardScaler()
scaler.fit(X_train.loc[:, continuous_features])

def transformation (data):
    continuous = pd.DataFrame(scaler.transform(data.loc[:, continuous_features]), columns=continuous_features, index=data.index)
    data = data.drop(continuous_features, axis=1)
    data = pd.concat([data, continuous], axis=1)
    return (data)

X_train = transformation(X_train)
X_test = transformation(X_test)

print(X_train.tail(5))
print('X_train.shape', X_train.shape, sep= ' ')
print('X_test.shape', X_test.shape, sep= ' ')
print('y_train.shape', y_train.shape, sep= ' ')
print('y_test.shape', y_test.shape, sep= ' ')
```

```python
def tsne_vizual (X_df):
    f, ax = plt.subplots(figsize=(10, 10))
    sns.set_context('poster', font_scale=0.6)
    cmap = sns.color_palette(['mediumseagreen', 'darkturquoise'], as_cmap=True)
    sns.scatterplot(x='t-SNE_component 1', y='t-SNE_component 2', hue='clas', data=X_df,
                    palette=cmap,
                    sizes=(10, 10),
                    edgecolor= 'dimgray')
    handles, labels = ax.get_legend_handles_labels()
    f.patch.set_facecolor(background)
    ax.legend(handles,
              ['0', '1'],
              scatterpoints=1, title = 'output', facecolor=background, framealpha=0)
    ax.text(-2, 22,
         't-SNE\n ____________________________________',
         ha='center', va = 'top',
         fontsize=16, fontweight='bold',
         color='black')
    ax.set_facecolor(background)

    return ()
```

```python
def learning_visual (model):
    color_list = ['black', 'darkred', 'saddlebrown',
                  'darkgoldenrod','olive', 'darkgreen',
                  'teal']
    from sklearn.metrics import roc_curve

    import matplotlib.gridspec as gridspec
    fig_cont = plt.figure(constrained_layout=True, figsize=(11,15))
    sns.set_context('poster', font_scale=0.6)
    spec2 = gridspec.GridSpec(ncols=2, nrows=9, figure=fig_cont)
    background = 'cornsilk'
    fig_cont.patch.set_facecolor(background)

    fig_title_ax = fig_cont.add_subplot(spec2[0, :])
    fig_title_ax.spines["bottom"].set_visible(False)
    fig_title_ax.spines["left"].set_visible(False)
    fig_title_ax.spines["top"].set_visible(False)
    fig_title_ax.spines["right"].set_visible(False)
    fig_title_ax.tick_params(left=False, bottom=False)
    fig_title_ax.set_yticklabels([])
    fig_title_ax.set_xticklabels([])
    fig_title_ax.set_facecolor(background)
    fig_title_ax.text(0.475 , 0.5,
                      '{}\n ____________________________________'.format(model),
                      horizontalalignment='center',
                      fontsize=20, fontweight='bold',
                      color='black')

    f_classification_report_ax = fig_cont.add_subplot(spec2[1:4, 0])
    y_predict = model.predict(X_test)
    label = classification_report(y_test, y_predict)
    f_classification_report_ax.text(0.95 , 0.5,
                                    '{}'.format(label),
                                    ha='right', va = 'center',
                                    fontsize=15,
                                    color='black')


    f_plot_roc_curve_ax = fig_cont.add_subplot(spec2[1:4, 1:2])
    y_pred_rf = model.predict_proba(X_test)[:, 1]
    fpr_rf, tpr_rf, _ = roc_curve(y_test, y_pred_rf)
    sns.lineplot(x=fpr_rf, y=tpr_rf, estimator=None, color = np.random.choice(color_list))
    sns.lineplot(x=[0, 1], y=[0, 1], color = 'black')
    f_plot_roc_curve_ax.set_facecolor(background)
    f_plot_roc_curve_ax.set_title('ROC', fontdict={'fontsize':14}, pad=10)
    f_plot_roc_curve_ax.set_xlabel('False positive rate', fontsize = 15)
    f_plot_roc_curve_ax.set_ylabel('True positive rate', fontsize = 15)


    f_ler_cur_ax = fig_cont.add_subplot(spec2[4:9, :])
    def learning_c(model):
        train_sizes, train_scores, test_scores = learning_curve(estimator=model,
                                                            X=X_train, y=y_train,
                                                            train_sizes=np.linspace(0.1, 1.0, 20),
                                                            cv=10,n_jobs=-1)
        return (train_sizes, train_scores, test_scores)

    learning_curve_point = learning_c(model)
    sns.lineplot(x = np.repeat(learning_curve_point[0], 10),
                 y = learning_curve_point[1].ravel(),
                 ci = 'sd', color = np.random.choice(color_list),  label = 'train')
    sns.lineplot(x = np.repeat(learning_curve_point[0], 10),
                 y = learning_curve_point[2].ravel(), sizes= 1,
                 ci = 'sd', color = np.random.choice(color_list), label = 'test')
    f_ler_cur_ax.set_xlabel('Data size', fontsize = 15)
    f_ler_cur_ax.set_ylabel('Accuarcy', fontsize = 15)
    f_ler_cur_ax.set_facecolor(background)
    f_ler_cur_ax.set_ylim([0, 1.1])
    f_ler_cur_ax.set_xlim([learning_curve_point[0][0], learning_curve_point[0][-1]])
    f_ler_cur_ax.set_title('Learning curve',fontdict={'fontsize':14}, pad=10)
    f_ler_cur_ax.legend(fontsize = 18, loc = 'lower left', facecolor=background, framealpha=0)


    for fig in [f_classification_report_ax]:

        fig.spines["bottom"].set_visible(False)
        fig.spines["left"].set_visible(False)
        fig.spines["top"].set_visible(False)
        fig.spines["right"].set_visible(False)
        fig.tick_params(left=False, bottom=False)
        fig.set_yticklabels([])
        fig.set_xticklabels([])
        fig.set_facecolor(background)
        fig.set_ylabel("")
        fig.tick_params(bottom=False)
    return ()
```
