### Importing required Libraries

```python
import pandas as pd
import numpy as np
import random
import warnings
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
%matplotlib inline
```

```python
d = pd.read_csv("/kaggle/input/iris-flower-dataset/IRIS.csv")
```

### Data Cleaning

```python
d["species"].replace({'Iris-setosa':'0', 'Iris-versicolor':'1', 'Iris-virginica':'2'}, inplace = True)
```

```python
d
```

### Data Visualization

```python
import seaborn as sns
sns.pairplot(d,hue = 'species')
```

### Feature and Lable set

```python
X = d.drop(['species'], axis = 1)
Y = d['species']
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

### Modeling

```python
from sklearn.linear_model import LogisticRegression
m = LogisticRegression()
m.fit(x_train, y_train)
```

### Accuracy

```python
m.score(x_test, y_test)
```

```python
from sklearn import metrics
print("Model Accuracy :", metrics.accuracy_score(y_test, m.predict(x_test)))
```

### Visualization with Confusion matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, m.predict(x_test))
cm
```

```python
import seaborn as sns
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot = True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
