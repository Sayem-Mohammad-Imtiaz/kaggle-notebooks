Pokedata'nın korelasyon analizi için pokedata.corr() kullanabiliriz. 
"Korelasyon analizi, iki değişken arasında ilişki olup olmadığını, eğer varsa bu ilişkinin ne yönde ve ne derecede olduğunu belirlemek için kullanılan bir istatistiksel yöntemdir. Örneğin; Sosyal medya araçlarının kullanımı ile Depresyon arasındaki ilişkinin araştırılmasında korelasyon analizi kullanılmaktadır. "
Değer 1 ya da 1e yakın ise pozitif bir ilişki, eğer -1 ve -1e yakın ise negatif bir ilişki vardır diyebiliriz. 



```python
print('Max Total: '+pokedata.loc[pokedata['Total'].idxmax()].Name+' / '+'TotalValue: '+str(pokedata['Total'].max()))
```

```python
print('Max Defense: '+pokedata.loc[pokedata['Defense'].idxmax()].Name+' / '+'DefenseValue: '+str(pokedata['Defense'].max()))
```

```python
print('Max Attack: '+pokedata.loc[pokedata['Attack'].idxmax()].Name+' / '+'AttackValue: '+str(pokedata['Attack'].max()))
```

```python
print('Min Total: '+pokedata.loc[pokedata['Total'].idxmin()].Name+' / '+'TotalValue: '+str(pokedata['Total'].min()))
```

```python
pokedata.corr()
```

Yukarıdaki tabloda special defence ile defence arasında bir ilişki olduğu ya da defansif özellikleri iyi olan bir pokemon'un ofansif özelliklerinin de iyi olması beklenebilir.  

Korelasyon haritasını yazdıralım. 

```python
f,ax = plt.subplots(figsize=(18,18))
sns.heatmap(pokedata.corr(), annot=True, linewidths= .5, fmt= '.1f',ax=ax)
plt.show()
```

```python
pokedata.Speed.plot(kind = 'line', color = 'g',label = 'Speed',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
pokedata.Defense.plot(color = 'r',label = 'Defense',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('x axis')              # label = name of label
plt.ylabel('y axis')
plt.title('Line Plot')            # title = title of plot
plt.show()
```

```python
# Scatter Plot 
# x = attack, y = defense
pokedata.plot(kind='scatter', x='Attack', y='Defense',alpha = 0.5,color = 'red')
plt.xlabel('Attack')              # label = name of label
plt.ylabel('Defence')
plt.title('Attack Defense Scatter Plot')            # title = title of plot
```

```python
# Histogram
# bins = number of bar in figure
pokedata.Speed.plot(kind = 'hist',bins = 50,figsize = (12,12))
plt.show()
```
