```python
df_stroke = pd.read_csv("../input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv")
```

```python
df_stroke.head()
```

```python
df_stroke = df_stroke.dropna()
```

```python
sns.countplot(x="heart_disease", data=df_stroke)
```

```python
sns.distplot(df_stroke['avg_glucose_level'])
```

```python
df_stroke = df_stroke.drop('ever_married', axis=1)
df_stroke = df_stroke.drop('id', axis=1)
```

```python
df_stroke = df_stroke.drop('gender',axis=1)
df_stroke = df_stroke.drop('work_type', axis=1)
df_stroke = df_stroke.drop('smoking_status', axis=1)
df_stroke = df_stroke.drop('Residence_type', axis=1)
```

```python
stroke = pd.get_dummies(df_stroke)
```

```python
y = stroke['stroke']
X = stroke.drop('stroke', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=11)
```

```python
print(X_train.shape, X_test.shape)
```

```python
stroke.head()
```

```python
xgb = XGBClassifier(n_estimatiors=500, learning_rate=0.1, max_depth=4)
xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)
```

```python
accuracy_score(y_test, xgb_pred)
```

```python
xgb = XGBClassifier(n_estimator=400, learning_rate=0.1, max_depth=3)
evals = [(X_test, y_test)]
xgb.fit(X_train, y_train, early_stopping_rounds=3, eval_metric="logloss", eval_set=evals, verbose=1)
```

```python
fig, ax = subplots()
plot_importance(xgb, ax=ax)
```
