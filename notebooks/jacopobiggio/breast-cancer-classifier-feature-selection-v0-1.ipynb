```python
b_data = pd.read_csv('../input/breast-cancer-wisconsin-data/data.csv', index_col = 'id')
b_data.shape
```

```python
b_data.head()
```

```python
b_clean = b_data.dropna(how = 'all', axis = 'columns')
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
b_clean.iloc[:, 0]= encoder.fit_transform(b_clean.iloc[:, 0]).astype('float64')
```

```python
plt.figure(figsize =(20,8))
cor = b_clean.corr()
sns.heatmap(data = cor, annot = True),
plt.show()
```

```python
cor_target = abs(cor['diagnosis'])
relevant = cor_target[cor_target>0.5]
relevant
var_final = relevant.drop('diagnosis')
var_final
```

```python
y= b_clean.diagnosis
X=b_clean.drop(['diagnosis'], axis = 1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
model = RandomForestClassifier(n_estimators = 100, random_state = 2)
scores = cross_val_score(model, X, y,
                         cv = 4,
                         scoring = 'accuracy')
scores.mean()
```
