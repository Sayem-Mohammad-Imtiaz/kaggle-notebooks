```python
# Create copy of index column named index currency
indexdata['Index Currency'] = indexdata['Index']

# Create dictionary of index names with their matching currency
index_to_currency = {}

for index, row in indexinfo.iterrows():
    index_to_currency_key = indexinfo.at[index, 'Index']
    index_to_currency_value = indexinfo.at[index, 'Currency']
    index_to_currency[index_to_currency_key] = index_to_currency_value

# Swap index codes within index currency column with their matching currency
indexdata['Index Currency'] = indexdata['Index Currency'].replace(index_to_currency)
```

```python
# Create copy of index currency column named index ratio
indexdata['Index Ratio'] = indexdata['Index Currency']

# Create dictionary of currency with their matching ratios
currency_exchange = {'USD' : 1}

for index, url in exchangerates.iterrows():
    currency_exchange_key = exchangerates.at[index, 'Currency']
    currency_exchange_value = exchangerates.at[index, 'Ratio']
    currency_exchange[currency_exchange_key] = currency_exchange_value

# Swap index currency withn index ratio column with their matching ratio
indexdata['Index Ratio'] = indexdata['Index Ratio'].replace(currency_exchange)
```

```python
indexdata['Open USD'] = indexdata['Open'] * indexdata['Index Ratio']
indexdata['High USD'] = indexdata['High'] * indexdata['Index Ratio']
indexdata['Low USD'] = indexdata['Low'] * indexdata['Index Ratio']
indexdata['Close USD'] = indexdata['Close'] * indexdata['Index Ratio']

# Export Table to .CSV
indexdata.to_csv('Export.csv', index = False)
```
