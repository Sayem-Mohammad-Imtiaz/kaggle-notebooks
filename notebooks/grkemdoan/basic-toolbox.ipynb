# Basic Visualization Data

```python
def lineplot(x,y):
    """ x : data 1
        y : data 2
        return : LinePlot
    """
    f,ax = plt.subplots(figsize = (9,5))
    x.plot(kind = "line",linestyle = ":",alpha = 0.5,linewidth = 3,color = "r",ax = ax,grid = True,)
    y.plot(kind = "line",linestyle = "-.",alpha = 0.5,linewidth = 3,color = "b",ax = ax,grid = True)
    plt.xlabel("Axis X")
    plt.ylabel("Axis Y")
    plt.show()
```

```python
liste = df.select_dtypes("float64","int64").columns
num = [0,1,2,3,4]
print([lineplot(df[liste[i]],df[liste[i+1]]) for i in num])
```

```python
def scatterPlot(x,y,hue = None):
    """ x : data 1
        y : data 2
        hue : If you want 3. dimension you will should use it
        return : ScatterPlot
    """
    f,ax = plt.subplots(figsize = (9,9))
    return sns.scatterplot(x,y,hue)
```

```python
liste = df.select_dtypes("float64","int64").columns
num = [0,1,2,3,4]
print([scatterPlot(df[liste[i]],df[liste[i+1]]) for i in num])
```

```python
def histogram(x):
    """ x : data 1

        return : Histogram
    """
    f,ax = plt.subplots(figsize = (9,9))
    return plt.hist(x,bins = 50)
```
