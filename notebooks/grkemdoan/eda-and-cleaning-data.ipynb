# Detection For Outliers Value & Suppress The Value

```python
for i,j in zip(df.dtypes.index,df.dtypes):
    if j == "int64" or j == "float64":
        if i == "Rank" or i == "Year":
            pass
        else:
            df.boxplot(i)
            plt.show()
```

```python
def lower_outliers(x):
    """
    x : Please get a variable 
    
    return: get lower_limit
    """
    Q1 = x.quantile(0.25)
    Q3 = x.quantile(0.75)
    IQR = 1.5 * (Q3 - Q1)
    
    lower_limit = Q1 - IQR
    
    return lower_limit
```

```python
def upper_outliers(x):
    
    """
    x : Please get a variable 
    
    return: get upper_limit
    """
    Q1 = x.quantile(0.25)
    Q3 = x.quantile(0.75)
    IQR = 1.5 * (Q3 - Q1)
    
    upper_limit = Q3 + IQR
    
    return upper_limit
```

```python
df_NaSales = df["NA_Sales"]
print("Lower limit : {}".format(lower_outliers(df_NaSales)))
print("Upper limit : {}".format(upper_outliers(df_NaSales)))
```

```python
for i,j in zip(df.dtypes.index,df.dtypes):
    if j == "int64" or j == "float64":
        if i == "Rank" or i == "Year":
            pass
        else:
            print("{} Lower limit : {}".format(i,lower_outliers(df[i])))
            print("{} Upper limit : {}".format(i,upper_outliers(df[i])))
```
