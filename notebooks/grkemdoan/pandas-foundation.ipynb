# Date Variable Updating

```python
year = []
month = []
day = []
for i in df["date"].index:
    year.append(df["date"][i].rsplit(df["date"][i][4:])[0])
    month.append(df["date"][i].rsplit(df["date"][i][6:])[0][4:])
    day.append(df["date"][i].rsplit(df["date"][i][8:])[0][6:])
```

```python
datetime = []
```

```python
for i,j,z in zip(year,month,day):
    datetime.append(i + "/" + j + "/" + z)

df["date"] = datetime
```
