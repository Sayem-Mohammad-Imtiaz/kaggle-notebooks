### Hello
### This is a mini tutorial on how to use LDA with Sklearn
### We'll be using the heart failure dataset, and using LDA to find the line that best separates classes.

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv")
```

```python
data.columns
```

I'll be using only two features here, but you can use the whole dataset if you like.  
Although you'll have to clean it and encode categorical features first.

```python
X=data.loc[:,['age','serum_creatinine']]
Y=data['DEATH_EVENT']
```

```python
LDA=LinearDiscriminantAnalysis()
```

```python
lda_data=LDA.fit_transform(X,Y)
```

What we get by applying that transformation is the projection of each data point (sample/line) on the LDA line.

The following are the projections of the first five samples.

```python
lda_data[:5]
```

You can add this 1D array now as a feature in your dataset. It almost always improves classification results.
