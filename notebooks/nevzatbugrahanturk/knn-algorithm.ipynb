```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../input/biomechanical-features-of-orthopedic-patients/column_2C_weka.csv")
data.head()
```

```python
data.rename(columns ={'class':'case'},inplace=True)
```

```python
A = data[data.case == "Abnormal"]
N = data[data.case == "Normal"]
```

```python
A.info()
```

# Scatter plot

```python

plt.scatter(A.lumbar_lordosis_angle,A.pelvic_radius,color="red",label="kotu",alpha=0.3)
plt.scatter(N.lumbar_lordosis_angle,N.pelvic_radius,color="green",label="iyi",alpha=0.3)
plt.xlabel("sacral_slope")
plt.ylabel("pelvic_radius")
plt.legend()
plt.show()
```

```python

data.case = [1 if each == "Normal" else 0 for each in data.case]
x_data = data.drop(["case"],axis = 1)
y = data.case.values
```

# Normalization

```python

x = (x_data - np.min(x_data)) / (np.max(x_data)-np.min(x_data))
```

```python
x
```

# Train Test Split

```python

from sklearn.model_selection import train_test_split
x_train, x_test , y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state = 1)
```

```python
# knn model
from sklearn.neighbors import KNeighborsClassifier
# find k value
score_list = []

for each in range(1,50):
    knn = KNeighborsClassifier(n_neighbors=each)
    knn.fit(x_train,y_train)
    score_list.append(knn.score(x_test,y_test))
    
plt.plot(range(1,50),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors = 21 ) #n_neighbors = k
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
```
