```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**The dataset contains cases from a study that was conducted between 1958 and 1970 
at the University of Chicago's Billings Hospital on the survival of patients who had 
undergone surgery for breast cancer.**

**The dataset contains the following Features :-**
*     Age Of the Patient at the time of Operation
*     Year of operation (In 1900's)
*     No. of positive axillary Nodes detected

    

**Objective : To predict Survival Status**

    1 = the patient survived 5 years or longer
    2 = the patient died within 5 year

```python
columns=['Age','Op_Year','Nodes','Status']
data=pd.read_csv("../input/habermans-survival-data-set/haberman.csv",header=None)
data.columns=columns
```

```python
data
```

```python
data.Status.value_counts()
# Majority of people tend to survive more than 5 yrs.
```

```python
plt.scatter(x=data.Age,y=data.Nodes)
plt.title("Age vs Nodes")
plt.xlabel("Age")
plt.ylabel("Nodes Detected")
plt.show()
# Majority Nodes are b/w 0 and 10
```

```python
plt.scatter(x=data.Age,y=data.Nodes)

plt.ylim(-1, 11)
plt.xlabel("Age")
plt.ylabel("Nodes Detected")
plt.show()
```

```python
# A closer look into 30-40 Age Gp
plt.scatter(x=data.Age,y=data.Status)
plt.xlim(29,41)
plt.show()
```

```python
plt.hist(data.Age,bins=50)
plt.title("Count of Age of Patients")
plt.show()
```

```python
plt.hist(data.Nodes,bins=50)
plt.title("Count of Axillary Nodes Detected")
plt.xlabel("No. of Nodes Detected")
plt.ylabel("Count of Patients")
plt.show()
```

```python
counts, bin_edges = np.histogram(data.Nodes, bins=50)
pdf = counts/(sum(counts))
```

```python
print(counts)
print(pdf)
print(bin_edges)
```

```python
# Probability Density Function
plt.plot(bin_edges[1:],pdf)
plt.show()
```

```python
# Cumulative Density Function
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf)
plt.show()
```

```python
sns.jointplot(x=data.Age,y=data.Nodes,data=data,kind="kde",hue=data.Status,height=8)
plt.ylim(-2,30)
plt.show()

# A kernel density estimate (KDE) plot is a method for visualizing the distribution 
# of observations in a dataset, analagous to a histogram. KDE represents the data using 
# a continuous probability density curve in one or more dimensions.
```

```python
sns.jointplot(x=data.Age,y=data.Nodes,data=data,kind="scatter",hue=data.Status,height=8)
plt.ylim(-2,30)
plt.show()
```
