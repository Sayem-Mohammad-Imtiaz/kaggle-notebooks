```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../input/breast-cancer-wisconsin-data/data.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})
```

### Plotting target column to check if data is equally divided. 

```python
sns.countplot(x="diagnosis", data=df)
```

```python
df.drop(columns=['id', 'Unnamed: 32'], inplace=True)
```

```python
plt.figure(figsize=(25,15))

matrix = np.triu(df.corr())
sns.heatmap(df.corr(), annot=True, mask=matrix, cmap="YlGnBu")
```

### From the heatmap we can see there are lot of features which are highly correlated, hence eliminating them to reduce **multicolinearity**.

```python
df.drop(columns=['radius_worst','perimeter_worst','area_worst','perimeter_se','area_se','perimeter_mean','area_mean','concave points_mean'], inplace=True)
```

```python
features = []
for col in df.columns:
    features.append(col)
    
features.remove('diagnosis')
target = ['diagnosis']
```

```python
from sklearn.model_selection import train_test_split

X = df[features]
y = df[target]

X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.fit_transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression()

log_reg.fit(X_train, y_train)
```

```python
print('Accuracy of logistic regression classifier on test set is : {:.2f}'.format(log_reg.score(X_test, y_test)))
```

```python
from sklearn.svm import SVC

clf = SVC()

clf.fit(X_train, y_train)
```

```python
print('Accuracy of svm classifier on test set is : {:.2f}'.format(clf.score(X_test, y_test)))
```

```python
from sklearn.metrics import classification_report, confusion_matrix

y_pred = log_reg.predict(X_test)

print("Confusion Matrix for Logistic Regression: ")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report for Logistic Regression: ")
print(classification_report(y_test, y_pred))
```

```python
y_pred = clf.predict(X_test)

print("Confusion Matrix for SVM Classifier: ")
print(confusion_matrix(y_test, y_pred))

print("\nClassification Report for SVM Classifier: ")
print(classification_report(y_test, y_pred))
```
