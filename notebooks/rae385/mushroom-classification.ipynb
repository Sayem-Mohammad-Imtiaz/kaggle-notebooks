```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("../input/mushrooms.csv")
data.head()
```

```python
features = data.iloc[:,1:]
features = pd.get_dummies(features).astype(float)  #one hot encoding
target = data.iloc[:,0]
features.head()
```

```python
X = features.values
y = target.values
```

Now because we are dealing with bunch of categorical data, many of which are binary data (true or false), decision trees will suffice as a well-rounded model for this problem.

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.30)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
```

```python
clf.fit(x_train, y_train)
```

```python
predictions = clf.predict(x_test)
print(accuracy_score(y_test,predictions))  # raw score
```

Let's test it out individually: we are going to pick out the last mushroom in the dataset.
