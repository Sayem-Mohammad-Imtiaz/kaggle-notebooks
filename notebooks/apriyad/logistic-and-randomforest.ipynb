```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.linear_model import LogisticRegression
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/kaggle/input/titanic/train_and_test2.csv')
```

```python
df.dropna(subset=['Embarked'],inplace=True)
```

```python
nunique_value=df.apply(pd.Series.nunique)
single_value=nunique_value[nunique_value==1].index.tolist()
print(single_value)
```

```python
df=df.drop(single_value,axis=1)
```

```python
df.shape
```

```python
x=df.drop(['survived'],axis=1)
```

```python
y=df['survived']
```

```python
x.shape,y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
x_train=x_train.drop(['Passengerid'],axis=1)
res=pd.DataFrame({'Passengerid': x_test['Passengerid']})
x_test=x_test.drop(['Passengerid'],axis=1)
```

```python
res.shape
```

```python
res.head()
```

Logistic Regression

```python
model_lr=LogisticRegression(max_iter=400)
model_lr.fit(x_train,y_train)
predict_lr=model_lr.predict(x_test)
accuracy_lr=accuracy_score(predict_lr,y_test)*100
print("Accuracy for Logistic Regression is {:.2f}".format(accuracy_lr))
```

Logistic Regression using Standard Scaler

```python
scaler=StandardScaler()
x_train_ss=scaler.fit_transform(x_train)
x_test_ss=scaler.fit_transform(x_test)

model_lr=LogisticRegression()
model_lr.fit(x_train_ss,y_train)
predict_lr=model_lr.predict(x_test_ss)
accuracy_lr=accuracy_score(predict_lr,y_test)*100
print("Accuracy for Logistic Regression using Standard Scaler is {:.2f}".format(accuracy_lr))
```

Logistic Regression using MinMax Scaler

```python
scaler=MinMaxScaler()
x_train_mm=scaler.fit_transform(x_train)
x_test_mm=scaler.fit_transform(x_test)
model=LogisticRegression()
model.fit(x_train_mm,y_train)
predict=model.predict(x_test_mm)
accuracy=accuracy_score(predict,y_test)*100
print("Accuracy for Logistic Regression using MinMax Scaler is {:.2f}".format(accuracy))
```

logistic Regression using MaxAbs Scaler

```python
scaler=MaxAbsScaler()
x_train_ma=scaler.fit_transform(x_train)
x_test_ma=scaler.fit_transform(x_test)
model=LogisticRegression()
model.fit(x_train_ma,y_train)
predict=model.predict(x_test_ma)
accuracy=accuracy_score(predict,y_test)*100
print("Accuracy for Logistic Regression using MaxAbs Scaler is {:.2f}".format(accuracy))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(max_depth=10)
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn.neural_network import MLPClassifier
model=MLPClassifier(solver='adam',activation='relu',max_iter=500,hidden_layer_sizes=(8,8,8))
model.fit(x_train,y_train)
mlp_ans=model.predict(x_test)
```

```python
accuracy= accuracy_score(mlp_ans,y_test)*100
print('Accuracy of MLP: {:.2f}'.format(accuracy))
```
