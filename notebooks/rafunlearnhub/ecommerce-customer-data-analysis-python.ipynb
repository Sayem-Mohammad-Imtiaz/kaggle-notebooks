```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
%matplotlib inline
```

```python
df = pd.read_csv('../input/ecommerce-events-history-in-cosmetics-shop/2019-Dec.csv')
df.head()
```

```python
df.describe()
```

```python
df1 = df.drop(['product_id', 'category_id', 'category_code', 'user_id', 'user_session'], axis=1)
df1.head()
```

```python
df1['event_type'].unique()
```

```python
event_map = {'view':2, 'cart':3, 'purchase':4, 'remove_from_cart':1}
df1['event_type'] = df1['event_type'].map(event_map)
df1.head()
```

```python
df1['row_num'] = np.arange(len(df1))
```

```python
datetimes = pd.to_datetime(df['event_time'])

df1['day'] = datetimes.dt.day

df1['month'] = datetimes.dt.month

df1.head()
```

```python
#{'view':2, 'cart':3, 'purchase':4, 'remove_from_cart':1}

purchases = 0; carts = 0; views = 0; removes = 0;

for i in range(0, len(df)):
    if df1['event_type'][i]==2:
        views+=1
    elif df1['event_type'][i]==3:
        carts+=1
    elif df1['event_type'][i]==4:
        purchases+=1
    elif df1['event_type'][i]==1:
        removes+=1

print("Views", views)
print("Carts", carts)
print("Purchases", purchases)
print("Removes", removes)
```

```python
pur1 = 0; pur5 = 0; pur10 = 0; pur15 = 0; pur20 = 0; pur25 = 0; pur30 = 0;
pric1 = 0; pric5 = 0; pric10 = 0; pric15 = 0; pric20 = 0; pric25 = 0; pric30 = 0;

for i in range(0, len(df)):
    
    if df1['day'][i]==1:
        pur1 += 1
        pric1 += df1['price'][i]
    
    elif df1['day'][i]==5:
        pur5 += 1
        pric5 += df1['price'][i]
    
    elif df1['day'][i]==10:
        pur10 += 1
        pric10 += df1['price'][i]
        
    elif df1['day'][i]==15:
        pur15 += 1
        pric15 += df1['price'][i]
        
    elif df1['day'][i]==20:
        pur20 += 1
        pric20 += df1['price'][i]
        
    elif df1['day'][i]==25:
        pur25 += 1
        pric25 += df1['price'][i]
        
    elif df1['day'][i]==30:
        pur30 += 1
        pric30 += df1['price'][i]

print(pur1)
print(pur5)
print(pur10)
print(pur15)
print(pur20)
print(pur25)
print(pur30)
print("SALES")
print(pric1)
print(pric5)
print(pric10)
print(pric15)
print(pric20)
print(pric25)
print(pric30)
```

```python
#{'view':2, 'cart':3, 'purchase':4, 'remove_from_cart':1}
def daily_vcpr(j):
    
    views1=0; carts1=0; purchases1=0; removes1=0;

    #j=1
    
    for i in range(0, len(df)):

        if df1['day'][i]==j:
            if df1['event_type'][i]==2:
                views1 +=1
            elif df1['event_type'][i]==3:
                carts1 +=1
            elif df1['event_type'][i]==4:
                purchases1 +=1
            elif df1['event_type'][i]==1:
                removes1 +=1
                
    cart_aband_rate = removes1 / carts1
    
    conversion_impact = purchases1 / carts1
    
    print("DAY: ", j)
    print("VIEWS: ",views1 );print("CARTS: ", carts1);print("PURCHASE: ", purchases1);print("REMOVES: ", removes1);
    
    print("CART ABANDONMENT RATE: ",cart_aband_rate); print("IMPACT OF CONVERSION: ",conversion_impact)
```
