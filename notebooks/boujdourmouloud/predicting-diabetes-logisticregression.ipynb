```python
# features and target
features = list(data.columns)[:-1]
target='Outcome'
X=data[features]
y=data[target]

# Split the Data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
# Check the shape of the data spliting
data_split = [X_train,X_test,y_train,y_test]
data_split_string = ['X_train','X_test','y_train','y_test']
for i,j in zip(data_split_string,data_split):
    print(f'The Shape of {i} is : {j.shape}')
```

```python
# Building the Model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
# Predictions
predictions = model.predict(X_test)
predictions
```

```python
# Model performence
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(predictions,y_test)
RMSE = (mse)**1/2
print(f'The value of mse is : {mse}')
print(f'The value of RMSE is : {RMSE}')
```

```python
# The accuarcy
accuarcy = model.score(X_test,y_test)
print(f'The value of accuarcy is : {accuarcy}')
```

```python
# The confusion matrix
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_test,predictions)
conf_matrix
```
