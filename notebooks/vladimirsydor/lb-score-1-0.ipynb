```python
import os

from os.path import join as pjoin

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

from sklearn.metrics import f1_score
```

# Read Data

```python
PATH2DATA = '../input/text-classification-int20h/'
os.listdir(PATH2DATA)
```

```python
train = pd.read_csv(pjoin(PATH2DATA, 'train.csv'))
test = pd.read_csv(pjoin(PATH2DATA, 'test.csv'))
sample_submission = pd.read_csv(pjoin(PATH2DATA, 'submission.csv'))
```

## Read IMDB data

```python
add_data = pd.read_csv('../input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv')
```

# Initial checks

```python
print(f"Train reviews intersection : {len(set(add_data.review) & set(train.review))}. Train len : {train.shape[0]}")
print(f"Train reviews intersection : {len(set(add_data.review) & set(test.review))}. Test len : {test.shape[0]}")
```

# Merge Train and IMDB

```python
train_add_data = train.merge(add_data, on='review', how='inner').drop_duplicates('review')

train_add_data = train_add_data.rename(columns={
    "sentiment_x":"true",
    "sentiment_y":"pred"
})

train_add_data['pred'] = train_add_data['pred'].map({
    "positive":1,
    "negative":0
})

print(f"Train F score: {f1_score(train_add_data.true, train_add_data.pred)}")
```

# Merge Test and IMDB

```python
test_add_data = test.merge(add_data, on='review', how='inner').drop_duplicates('review')

test_add_data['sentiment'] = test_add_data['sentiment'].map({
    "positive":1,
    "negative":0
})

test_add_data
```

```python
test_add_data.drop(columns='review').to_csv('submission.csv', index=False)
```
