# Problem definition and task
    
    The given dataset is about the different species of Mushroom 
    and based on the different properties of mushroom which is given
    as categorical features. We will further explore the data in the notebook.
    we have to predict or classify whether a particular species of mushroom is 
    ediable or poisonous. so it clear that it is classification task and we are
    going to use LogisticRegression Classifier and Random Forests Classifier 
    algorithms for this problem.
    
    Let's get started....!!

```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
# Reading the data
df = pd.read_csv('../input/mushroom-classification/mushrooms.csv')
df.head()
```

```python
# Let's check the size of the dataset
df.shape
```

```python
# See if there are any null values
df.isnull().sum()
```

```python
# See count of unique values in each column
df.nunique()
```

 # Preprocessing
     **Great, there are no null or missing values in any of the column but all the columns
     consist of categorical values so we need to convert them into numerical values  **

```python
encoder = LabelEncoder()
```

```python
# Assigning each categorical class in column to numerical value
mapping = list()
for i in range(len(df.columns)):
    df[df.columns[i]] = encoder.fit_transform(df[df.columns[i]])
    mapping_dict = {after_encoding: before_encoding for after_encoding, before_encoding in enumerate(encoder.classes_)}
    mapping.append(mapping_dict)
mapping
```

```python
# let see the data now
df.head()
```

```python
# Now decide our features and label(targeted column)
X = df.drop("class", axis=1)
y = df['class']
```

```python
# splitting the dataset into training and testing (30 %) set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
# Now let's standarize the data so that all the column values fall in the same range.
scaler = StandardScaler()
X_train = pd.DataFrame(scaler.fit_transform(X_train), columns = X_train.columns)
X_test = pd.DataFrame(scaler.transform(X_test), columns = X_test.columns)
```

```python
X_train.head()
```

```python
type(X_train.describe())
```

```python
# see the standard deviation of all the columns becomes around 1.
X_train.describe().T['std']
```

# Training

```python
LR = LogisticRegression(solver='lbfgs')
rfc = RandomForestClassifier(n_estimators = 10)  # n_estimators that is num of trees are taken as 10 intitally.
```

```python
# fitting data to models
LR.fit(X_train, y_train)
rfc.fit(X_train, y_train)

# predicting
LR_pred = LR.predict(X_test)
rfc_pred = rfc.predict(X_test)
```

```python
print("-"*10 + "The performance parameters for LR model" + "-"*10, "\n")
print("The Confusion_matrix is : \n", confusion_matrix(y_test, LR_pred), "\n\n", "-"*25)
print("The classificaton_report is : \n", classification_report(y_test, LR_pred), "\n", "-"*25)
print("The accuracy of the model is :", accuracy_score(y_test, LR_pred), '\n\n\n')


print("-"*10 + "The performance parameters for Random Forests model" + "-"*10, "\n")
print("The Confusion_matrix is : \n", confusion_matrix(y_test, rfc_pred), "\n\n", "-"*25)
print("The classificaton_report is : \n", classification_report(y_test, rfc_pred), "\n", "-"*25)
print("The accuracy of the model is :", accuracy_score(y_test, rfc_pred))
```
