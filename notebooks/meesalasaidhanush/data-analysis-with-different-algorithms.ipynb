# building the model

# support vector machine

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn import svm
```

```python
sv=svm.SVC(gamma=0.01,C=1,kernel='linear')
sv.fit(x_train,y_train)
pred=sv.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,pred))
```

# logistic regression

```python
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
pre=lr.predict(x_test)
```

```python
print(classification_report(y_test,pre))
```

# xgboost algorithm

```python
import xgboost
```

```python
xbg=xgboost.XGBClassifier()
```

```python
xbg.fit(x_train,y_train)
pr=xbg.predict(x_test)
```

```python
print(classification_report(y_test,pr))
```

# decision tree and random forest

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
tree=DecisionTreeClassifier()
```

```python
tree.fit(x_train,y_train)
p=tree.predict(x_test)
```

```python
print(classification_report(y_test,p))
```

```python
forest=RandomForestClassifier()
```

```python
forest.fit(x_train,y_train)
pp=forest.predict(x_test)
```

```python
print(classification_report(y_test,pp))
```

# from the above classification models we can conclude that 'support vector machine' gives us the highest accuracy
