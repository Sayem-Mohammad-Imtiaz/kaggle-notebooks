 # data preprocessing 

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
l=LabelEncoder()
```

```python
df['Genre']=l.fit_transform(df['Genre'])
```

```python
df['Publisher']=l.fit_transform(df['Publisher'])
```

```python
df.head()
```

# feature selection

```python
cor=df.corr()
cor
```

```python
sns.heatmap(cor)
```

```python
x=df.iloc[:,[4,5,6,7,8,9]]
y=df.iloc[:,[10]]
```

```python
y
```

# data scaling

```python
sc=StandardScaler()
```

```python
x=sc.fit_transform(x)
```

```python
x
```

# models creation

## Regression model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
# lets start with the basic regression model i.e multiple regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
pred
```

```python
print(mean_squared_error(pred,y_test))
```

```python
sns.distplot(y_test-pred)
```

## xgboost model

```python
import xgboost
```

```python
xgb=xgboost.XGBRegressor()
xgb.fit(x_train,y_train)
```

```python
pred1=xgb.predict(x_test)
```

```python
pred1
```

```python
print(mean_squared_error(pred1,y_test))
```

## decision tree model

```python
from sklearn.tree import DecisionTreeRegressor
tree= DecisionTreeRegressor()
```

```python
tree.fit(x_train,y_train)
```

```python
pred2=tree.predict(x_test)
pred2
```

```python
print(mean_squared_error(pred2,y_test))
```

#### from the  above mentioned few models multiple regression is giving us least error
