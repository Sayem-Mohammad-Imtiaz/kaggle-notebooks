```python
df.drop(["Unnamed: 32"], axis = 1, inplace = True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
import random
import time
from sklearn.metrics import confusion_matrix,classification_report
```

```python
x=df.drop(["diagnosis"],axis=1).values
y=df["diagnosis"].values
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.3)
```

```python
model_1=RandomForestClassifier(random_state=42, bootstrap=True, criterion='gini', max_depth=4, max_features='log2')
```

```python
model_1.fit(x_train,y_train)
```

```python
predicted_1=model_1.predict(x_test)
```

```python
print("Training Acuracy:",model_1.score(x_train,y_train)*100)
print("Testing Acuracy:",model_1.score(x_test,y_test)*100)
```

```python
print("\n Classification_report: \n",classification_report(y_test,predicted_1))
```

```python
my_submission= pd.DataFrame({'diagnosis':predicted_1})
my_submission.to_csv('submission.csv',index=False)
```
