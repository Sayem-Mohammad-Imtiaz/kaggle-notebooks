```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x=df.drop("Salary",axis=1).values
y=df["Salary"].values
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3)
```

```python
lr=LinearRegression()
model=lr.fit(x_train,y_train)
predict=model.predict(x_test)
```

### Evaluation Metrics
1.  MAE(Mean Absolute Error) calculates the absolute difference between actual and predicted values.
   * sum all the errors and divide them by a total number of observations.
   * aim to get a minimum MAE because this is a loss.
   * robust to outliers.
   * MAE is not differentiable so we have to apply various optimizers like Gradient descent which can be differentiable
2. Mean Squared Error(MSE) is the squared difference between actual and predicted value.
  *  It is not Robust to outliers.
3. Root Mean Squared Error(RMSE) is square root of mean squared error.
  * it is not that robust to outliers
4. R2 score is a metric that tells the performance of your model.
  * R2 calculates how must regression line is better than a mean line.
  * R2 is also known as Coefficient of Determination or sometimes also known as Goodness of fit.
5. Adjusted R Squared 
  * adj_r2_score = 1 - ((1-r2)*(n-1)/(n-p-1))

```python
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.metrics import mean_squared_error
MAE=mean_absolute_error(y_test,predict)
print("MAE=",MAE,"\n")
MSE=mean_squared_error(y_test,predict)
print("MSE=",MSE,"\n") 
RMSE=np.sqrt(mean_squared_error(y_test,predict))
print("RMSE=",RMSE,"\n")
r2=r2_score(y_test,predict)
print("r2=",r2,"\n")
p=1 # no of independent variable
n=len(df)
adj_r2= 1 - ((1-r2)*(n-1)/(n-p-1))
print("adj_r2=",adj_r2,"\n") 

```

```python
print("Coefficient=",model.coef_)
print("\n")
print("Intercept=",model.intercept_)
```

```python
plt.scatter(x_test,y_test, color = 'red')
plt.plot(x_test, model.coef_*x_test + model.intercept_, '-b')
plt.title("Testing Set")
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.show()
```
