## Separating Dependent and Independent Variables

With the help of EDA we come to know that this is a classification problem in which we have independent and dependent variables, based on our independednt variables we will classifing or predicting our output. Assigning "X" as Independent Variable and "y" as Dependent or Target Variable

```python
X = data.drop(['left'], axis=1)
y = data.left
```

```python
X.head()
```

```python
y.head()
```

```python
X = X.drop(['last_evaluation', 'Work_accident'], axis='columns')
X.head()
```

Here we have our X for building our model, but before proceeding ahead we know that while building Machine Learning Model our dataset value should be numeric. In our dataset we still have categorical columns, so our duty here is to convert non-numeric columns to numeric.

There are various concepts for converting categorical columns to numeric:
1. One Hot Encoding
2. Label Encoder
3. Using Pandas Liberary
4. Manually allocating numeric value to each entry.

I am going to use LabelEncoder which is function of scikit Learn. Using labelencoder will convert it by itself within same column.

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

In the above cell we initializes our labelencoder with valiable le. Next job is to fit the encoder and transform columns

```python
X['Department'] = le.fit_transform(X['Department'])
X['salary'] = le.fit_transform(X['salary'])
X.head()
```

Our Dataset is converted into numeric values. Everything is converted as we needed it to be before building or training model.

We need to check our dataset whether it is imbalanced or not

```python
from collections import Counter
Counter(y)
```

```python
imb = y.value_counts().head()[0]/len(y)
print(imb)
```

# Building Model

## Splitting Dataset into Train and Test
Splitting Dataset into two part train and test, the motive is simple that we will be building or training our model on training set and with the help of test set we will check our model performance in terms of accuracy.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
```

We are ready to train our model. Here we will be building our model and checking it's performance

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()

from sklearn.model_selection import cross_val_score
acc=cross_val_score(estimator=rf,X=X,y=y,cv=10,scoring='accuracy')
acc.mean()
```

```python
rf.fit(X_train,y_train)
```

```python
# Predicting our Model accuracy
y_pred = rf.predict(X_test)

from sklearn.metrics import accuracy_score
print(f"Random Forest ACC: {accuracy_score(y_test, y_pred)*100}")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
