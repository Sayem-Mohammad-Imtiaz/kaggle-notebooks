# Dogecoin price Analysis and Prediction

#### This is Dogecoin price Prediction and Analysis program using python. The price is in USD.

```python
#import all the libraries required 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
%matplotlib inline
sns.set_style('darkgrid')
```

```python
#store the price data into a dataframe 
data=pd.read_csv('../input/dogecoin-historical-data/DOGE-USD.csv')
```

```python
data=data.set_index(pd.DatetimeIndex(data['Date'].values))
```

```python
#lets visualize and analyze the data 
```

```python
fig=go.Figure(
    data=[go.Candlestick(
    x=data['Date'],
    close=data['Close'],
    open=data['Open'],
    high=data['High'],
    low=data['Low'],
    increasing_line_color='green',
    decreasing_line_color='red'    
    )])
fig.show()
```

```python
#lets predict the price 
```

```python
#We use only Close column to predict the price ,So lets use only Close value 
```

```python
df=data[['Close']]
```

```python
#lets define a variable to store number of days to predict
no_of_days=1 #for example
```

```python
#create a new column prediction
df['prediction']=df[['Close']].shift(-no_of_days)
```

```python
#this is the function to clean data without having nan or infinity values
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    #df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(1)
    return df[indices_to_keep].astype(np.float64)
```

```python
df=clean_dataset(df)
df
```

```python
X=np.array(df.drop(['prediction'],axis=1))
#we have to remove n+1 of data where n is the no_of_days
X=X[:len(df)-no_of_days-1]
```

```python
X
```

```python
y=np.array(df['prediction'])
#We have to get all the values of y execpt last n+1
y=y[:len(df) -no_of_days -1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X, y, test_size=0.33)
```

```python
from sklearn.ensemble import RandomForestRegressor
rnd=RandomForestRegressor(n_estimators=12,random_state=507)
```

```python
rnd.fit(X_train,y_train)
```

```python
rnd.score(X_test,y_test) #score of test set compared to the trainin 
```

```python
#lets compare the predicted values and the actual values 
predict=rnd.predict(X_test)
print(predict)
print(" ")
#actual values
print(y_test)
```

```python
#lets validate the model 
#lets have the dataset the contain all the data except for n rows
temp=df[:-no_of_days]
x_val=temp.tail(no_of_days)['Close'][0]
x_val
```

```python
predictions=rnd.predict([[x_val]])
```

```python
#lets print the value of doge upto n number of days
print("The price of  Dogecoin is ",no_of_days,"days is to be",predictions)
#The actual value is 
print('The actual price is ',temp.tail(no_of_days)['prediction'][0])#as we deleted the n row while cleaning 
```
