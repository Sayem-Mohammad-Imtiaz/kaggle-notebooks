```python
# Reading from csv file

data = pd.read_csv("../input/biomechanical-features-of-orthopedic-patients/column_2C_weka.csv", sep=",")

y_data = data["class"].values
x_data = data.drop("class", axis=1)

y = [0 if each == "Normal" else 1 for each in y_data]
```

```python
# Normalization

x = (x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data))
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)
```

```python
# KNN

from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
predict = knn.predict(x_test)
```

```python
# Score

print("KNN Score with {} neighbors: {}".format(5, knn.score(x_test, y_test)))
```

```python
# Finding a K value

import matplotlib.pyplot as plt

score_list = []
for each in range(1, 15):
    knn2 = KNeighborsClassifier(n_neighbors=each)
    knn2.fit(x_train, y_train)
    score_list.append(knn2.score(x_test, y_test))

plt.plot(range(1, 15), score_list)
plt.xlabel("K Neighbors")
plt.ylabel("Accuracy")

print("Optimal Score: {}".format(np.max(score_list)))
print("Optimal K value: {}".format(score_list.index(np.max(score_list))))
```
