Ahora que exploramos uno a uno los dataset podemos comenzar a realizarnos las preguntas necesarias para extraer la informacion mas relevante de estos dataset para tener un analisis detallado de la informacion que contienen los mismos.

### 1. QUE EQUIPO DE LA NBA HA GANADO MAS VECES?

```python
data_merge= pd.merge(left=ranking,right=teams,left_on="TEAM_ID",right_on="TEAM_ID")
wins= data_merge.groupby(by="NICKNAME").mean().round()
wins.sort_values("W",ascending=False,inplace=True)
```

```python
fg1=px.histogram(data_frame=wins ,x=wins.index,y=["W"],title="PROMEDIO DE PARTIDOS GANADOS POR EQUIPO",labels={"NICKNAME":"Nombre Equipo",
                                                                               "value":"Partidos Ganados"})
fg1.show()
```

En la imagen anterior podemos observar como los Spurs de San Antonio es el equipo mas ganador durante la fecha establecida en este dataset que va desde el año 2003 hasta el año 2021 seguido de los Rockets de Houston

Ahora podemos hacernos la pregunta contraria

### 2. QUE EQUIPO DE LA NBA HA PERDIDO MAS VECES?

```python
wins.sort_values("L",ascending=False,inplace=True)

fg2=px.histogram(data_frame=wins ,x=wins.index,y=["L"],title="PROMEDIO DE PARTIDOS PERDIDOS POR EQUIPO",labels={"NICKNAME":"Nombre Equipo",
                                                                               "value":"Partidos Perdidos"})
fg2.show()
```

Ahora en este grafico podemos observar que equipos han sido lo mas perdedores durante este tiempo en este caso vemos una clara relacion inversa al grafico anterior en este podemos ver que los lideran los Timberwolves de Minnesota y los Knicks de Nueva York

### 3. CANTIDAD DE JUEGOS GANADOS POR EQUIPO EN CONDICION DE LOCAL?

```python
data_super= pd.merge(left=teams,right=games,left_on="TEAM_ID",right_on="HOME_TEAM_ID")
team_wins= data_super.groupby("NICKNAME").sum()
team_wins.sort_values("HOME_TEAM_WINS",ascending=False,inplace=True)
```

```python
plt.figure(figsize=(18,10))
sns.barplot(x=team_wins.index,y=team_wins["HOME_TEAM_WINS"],)
plt.xticks(rotation=90)
plt.title("EQUIPOS MAS GANADORES EN CASA",fontsize=18,fontweight="bold")
plt.xlabel("Nombre Equipo",fontsize=14)
plt.ylabel("Cantidad de Partidos Ganados",fontsize=14)
plt.show()
```

En este grafico podemos identificar como los miami heats suben en el ranking de los equipos mas ganadores al analizar los juegos ganados como locales y los san antonio spurs se mantienen como lider de equipo mas ganador

### 4. ANALIZANDO LAS ESTADISTICAS DE LOS JUGADORES?

```python
jugadores= details.groupby("PLAYER_NAME").sum()
```

```python
jugadores.head()
```

```python
jugadores.columns
```

```python
data_player= jugadores[["PTS","REB","AST","STL","BLK"]]
```

```python
data_player.head()
```

```python
data_player.rename(columns={"PTS":"Puntos","REB":"Rebotes","AST":"Asistencias","STL":"Robos","BLK":"Bloqueos"},inplace=True)
```

```python
jugadores.rename(columns={"FGM":"Tiros Anotados","FGA":"Tiros Intentados",
                          "FG3M":"Tiro 3 Pts Anotados","FG3A":"Tiro 3 Pts Intentados","FTM":"Tiros Libres Anotados",
                         "FTA":"Tiros Libres Intentados"},inplace=True)
```

```python
jugadores.columns
```

```python
import plotly.graph_objects as go
```

```python
def Jugador_comparativo(player1,player2,player3):
    
    data= (data_player.index==player1)|(data_player.index==player2)|(data_player.index==player3)
    
    transpolar= data_player[data].T
    categories = transpolar.index
    
    c_layout= dict(title="GRAFICO ESTADISTICO JUGADOR")
    
    fig = go.Figure(layout=c_layout)
    
    fig.add_trace(go.Scatterpolar(r=transpolar[player1],theta=categories,fill="toself",name=player1))
    
    fig.add_trace(go.Scatterpolar(r=transpolar[player2],theta=categories,fill="toself",name=player2))
    
    fig.add_trace(go.Scatterpolar(r=transpolar[player3],theta=categories,fill="toself",name=player3))
    
    columns= [ 'Tiros Anotados', 'Tiros Intentados','Tiro 3 Pts Anotados', 'Tiro 3 Pts Intentados',
       'Tiros Libres Anotados', 'Tiros Libres Intentados']
    
    fig1= px.bar(jugadores[data][columns],title="OTRAS ESTADISTICAS",labels={"PLAYER_NAME":"Nombre Jugador","value":"Cantidad"})
    
        
    fig.show()
    fig1.show()
    
    
```

```python
Jugador_comparativo("Kevin Durant","LeBron James","Kobe Bryant")
```
