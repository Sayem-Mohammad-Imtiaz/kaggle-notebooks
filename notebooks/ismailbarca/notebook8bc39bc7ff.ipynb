```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split

```

```python
housing_num=housing.drop('ocean_proximity',axis=1)
cat=["ocean_proximity"]
num_feat=list(housing_num)
num_feat
```

```python
category=housing.ocean_proximity
encode=OneHotEncoder()
category.value_counts().index
```

```python
column_trans=ColumnTransformer([
    ("num",SimpleImputer(strategy="median"),num_feat),
    ("cat",OneHotEncoder(),cat)
])

house_prepared=column_trans.fit_transform(housing)
```

```python
house_prepared=pd.DataFrame(house_prepared,columns=['longitude',
 'latitude',
 'housing_median_age',
 'total_rooms',
 'total_bedrooms',
 'population',
 'households',
 'median_income',
 'median_house_value','<1H OCEAN', 'INLAND', 'NEAR OCEAN', 'NEAR BAY', 'ISLAND'])
```
