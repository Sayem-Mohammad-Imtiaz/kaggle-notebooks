# Customer Profile 


I want to show the correlation between Age and income and see the separated plots at the univarante plots.

Also, we can see it is right skewed. Indicate that our main focus should be on on young adults between middle 20's to 30 years old.

```python
plt.figure(figsize=(15,8)) #set the image size 
sns.distplot(fitnessData.Age, kde=True); #Univariate product usage per age.
plt.axvline(x=fitnessData.Age.mean(), linewidth=1, color='blue', label="mean", alpha=0.8); #show a green line where in the plot should be the mean
plt.axvline(x=fitnessData.Age.median(), linewidth=1, color='green', label="median", alpha=0.8); #show a blue line where in the plot should be the median

plt.legend(["mean", "median"]); #the plot legend for eman and meadian
```

The univariate plot by income. 

```python
plt.figure(figsize=(15,8)) #set the image size 
sns.distplot(x=fitnessData.Income) #Univariate product usage per income.
plt.show();
```

**Charactiristic of a customer of the different products**

```python
sns.catplot(x='Product', y='Age', hue='Gender', kind='bar', data=fitnessData);
#Seaborn catplot. product on y axel and the age for each customer gender on x axel
```

On This plot we can notice that the product TM195 has the highest count while the product TM798 are usaged more frequently 

```python
sns.displot(x='Usage', hue='Product', kind='kde', data=fitnessData);
#Uni-variate of the products by usage
```

Usage vs Income.  People with hightest income use more than the avarage people. 

```python
sns.jointplot(x='Usage', y='Income', data=fitnessData);# Seaborn joinplot. product on y axel and the age for each customer gender on x axel
```

Using the swarmplot we can visualize a multivariate plot. Usage vs Income vs Fitness. 

```python
plt.figure(figsize=(15,8)) #set the image size 
sns.swarmplot(data = fitnessData, x='Usage', y='Income', hue='Fitness'); #seaborn swarmplot 
```

```python
sns.pairplot(fitnessData,  hue="Product"); #seaborn pairplot to show the relationship in a dataset. 
```

The heatmap show the correlation among the dataset.

```python
plt.figure(figsize=(10,7)) #set the image size 
sns.heatmap(fitnessData.corr(), annot=True); #seaborn heatmap with the correlation amount the data in the dataset. use the parameter annot to display annotation for each square.
```

This plot show the product consumption by fitness category and Income 

```python
plt.figure(figsize=(15,8)) #set the image size 
ax = sns.barplot(x='Fitness', y='Income', data = fitnessData, hue='Product') #barplot showing the Fitness vs income for each product. 
```

On This Point Plot we can have the visualization by product, the fitness and income for all data

```python
plt.figure(figsize=(10,5)) #set the image size 
sns.pointplot(x='Fitness', y='Usage', hue='Product', data=fitnessData) #The pointplot to show how change the produt usage for each fitness session by product.

plt.show()
```
