```python
import pycaret
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import classification module 
from pycaret.classification import *
```

```python
#intialize the setup (in Notebook env)
exp_clf = setup(df1, target = 'class')
```

```python
# return best model
best = compare_models()
```

```python
# train a decision tree model with default parameters
lr = create_model('lr')
```

```python
# tune hyperparameters of logistic Regression
tuned_lr = tune_model(lr)
```

```python
# tune hyperparameters to optimize AUC
tuned_lr = tune_model(lr, optimize = 'AUC')
```

```python
# ensemble logistic Regression model with 'Bagging'
bagged_lr = ensemble_model(lr)
```

```python
# ensemble logistic Regression model with 'Boosting'
boosted_lr = ensemble_model(lr,method='Boosting')
```

```python
#plot a model
plot_model(lr,plot='confusion_matrix')
```

```python
plot_model(lr,plot='class_report')
```

```python
plot_model(lr,plot='feature')
```

```python
plot_model(lr,plot='parameter')
```

```python
# train a voting classifier dynamically
blender_top5 = blend_models(compare_models(n_select = 5))
```

```python
# stack trained models dynamically
top5 = compare_models(n_select = 5)
stacked_models = stack_models(estimator_list = top5[1:], meta_model = top5[0])
```
