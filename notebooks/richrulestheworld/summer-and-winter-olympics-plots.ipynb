```python
import numpy as np
import pandas as pd

# For data visualization 
import seaborn as sns 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
summer=pd.read_csv('../input/olympic-games/summer.csv')
```

```python
summer.head()
```

```python
summer.info()
```

```python
summer.isnull().sum()
```

```python
summer = summer.drop_duplicates()
summer.shape
```

```python
summer.nunique()
```

```python
male_p = summer['Gender'].value_counts()[0]
female_p = summer['Gender'].value_counts()[1]
total_p = male_p + female_p

gender_pchart = pd.DataFrame({'Gender': [male_p, female_p]}, index=['Male', 'Female'])
gender_pchart.plot.pie(y='Gender', figsize=(10, 10))

print("Male Participants %: {0:.2f}".format(male_p/total_p * 100))
print("Female Participants %: {0:.2f}".format(female_p/total_p * 100))
```

```python
summer['Sport'].value_counts().head(10).plot.barh(figsize=(10,10)).invert_yaxis()
plt.title('Most Popular Sports (Summer Olympics)', size = 20,)
```

```python
summer['Sport'].value_counts().tail(10).plot.barh(figsize=(10,10)).invert_yaxis()
plt.title('Least Popular Sports (Summer Olympics)', size = 20,)
```

```python

plt.figure(figsize=(20,10), edgecolor='black', linewidth=3)
male_winners = sns.countplot(x='Year', data= summer[summer.Gender == 'Men'], edgecolor=sns.color_palette("dark", 3))
plt.title('Male winners throughout the years (Summer)', size = 20,)
```

```python

#Setting plot size and edgecolors
plt.figure(figsize=(20,10), edgecolor='black', linewidth=3)

#Creating plot
female_winners = sns.countplot(x='Year', data= summer[summer.Gender == 'Women'], edgecolor=sns.color_palette("dark", 3))

    
#manipulating title visuals
plt.title('Female winners throughout the years (Summer)', size = 20,)
```

```python
medals_country=summer.groupby(['Country','Medal'])['Gender'].count().reset_index().sort_values(by='Gender',ascending=False)
medals_country=medals_country.pivot('Country','Medal','Gender').fillna(0)
top=medals_country.sort_values(by='Gold',ascending=False)[:10]
top.plot.barh(width=0.8,color=['#CD7F32','#FFDF00','#D3D3D3'], figsize=(10,10)).invert_yaxis()
plt.title('Medals Distribution Of Top 10 Countries (Summer Olympics)')
plt.show()
```

```python
medals_country=summer.groupby(['Discipline','Medal'])['Gender'].count().reset_index().sort_values(by='Gender',ascending=False)
medals_country=medals_country.pivot('Discipline','Medal','Gender').fillna(0)
top=medals_country.sort_values(by='Gold',ascending=False)[:10]
top.plot.barh(width=0.8,color=['#CD7F32','#FFDF00','#D3D3D3'], figsize=(10,10)).invert_yaxis()
plt.title('Medals Distribution Of Top 10 Sports (Summer Olympics)')
plt.show()
```

```python


test1=summer.groupby(['Country','Year'])['Medal'].count().reset_index()
test1=test1[test1['Country'].isin(summer['Country'].value_counts()[:10].index)]
test1=test1.pivot('Year','Country','Medal')
test1.plot()
fig=plt.gcf()
fig.set_size_inches(18,8)
plt.title('Medals By Years By Country (Summer Olympics)')
plt.show()

```

```python
summer.loc[summer['Discipline'].str.contains('Wrestling'),'Discipline']='Wrestling'
summer.loc[summer['Discipline'].str.contains('Weightlifting'),'Discipline']='Weightlifting'
test=summer[summer['Country'].isin(summer['Country'].value_counts()[:10].index)]
test=test[test['Discipline'].isin(summer['Discipline'].value_counts()[:10].index)]
test=test.groupby(['Country','Discipline'])['Sport'].count().reset_index()
test=test.pivot('Discipline','Country','Sport')
sns.heatmap(test,annot=True,fmt='2.0f')
fig=plt.gcf()
fig.set_size_inches(8,6)
plt.show()
```
