```python
clf = tree.DecisionTreeClassifier(criterion="entropy")
clf = clf.fit(train[['Pregnancies', 'Glucose', 'BloodPressure', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']], train['Outcome'])
```
