```python
#poverty rate of each state
#Her bir eyaletin yoksulluk oranı
percentage_people_below_poverty_level.poverty_rate.replace(["-"],0.0,inplace=True)

percentage_people_below_poverty_level.poverty_rate = percentage_people_below_poverty_level.poverty_rate.astype(float)

area_list = list(percentage_people_below_poverty_level["Geographic Area"].unique())
#bunun eyaletlerini al ve içinde unique olan değerleri bul dedim
#ardından bunun liste çevir dedim ve area_liste depoladım

area_poverty_ratio=[]
for i in area_list:
   
    x = percentage_people_below_poverty_level[percentage_people_below_poverty_level["Geographic Area"]==i]
    #datamın içinde area_liste kayıt etigim unique eyaletlerimi her birini tek tek bul diyorum
    
    area_poverty_rate = sum(x.poverty_rate)/len(x)
    # bunu bulduktan sonra benım poverty_rate ortalamasını aldiyorum
    #bunu area_poverty_ratio içinede saklıyorum
    
    area_poverty_ratio.append(area_poverty_rate)
    #daha sonrada bu variable daha onceden yaratıgım area_poverty_ratio listenın içine aktarıyorum

    data = pd.DataFrame({"area_list":area_list, "area_poverty_ratio":area_poverty_ratio})
#bırtane data olusturuyorum bu datamın columns bir tanesi area_list digeride area_poverty_ratio olsun dıyroum

new_index = (data["area_poverty_ratio"].sort_values(ascending=False)).index.values
#daha sonra bunu sort edıyorum ascending=False bunu yazmamın sebebı azalan sekılde olsun demek
#eger ben bunu true yapsaydım artan sekılde olack grafıgımız.Biz en fakir eyaleti istiyoruz

sorted_data = data.reindex(new_index)
#ardından sort edıyorum yukarda datamın ındexlerını cekmıstım artık bunu
#datamın yenı ındexi yapıyorum

#visualization (artık burda gorselestırme kısmına gecıyoruz)
plt.figure(figsize=(15,10))
#benım ıcın bır figür belirle ve bunun boyutu 15,10 yap diyorum

sns.barplot(x=sorted_data["area_list"],y=sorted_data["area_poverty_ratio"])
#sns benım seaborn numun kısaltması
#bu benım bar plotum x eksenim sorted_data["area_list"] eyalet listem y eksenimde sorted_data["area_poverty_ratio"])
#yani fakirlik oranı 

plt.xticks(rotation=90)#45de yapabılırız
# x eksemınde olan eyalet isimlerimi 90 derecelik açıyla koy diyorum

plt.xlabel("States")
plt.ylabel("Poverty Rate")
#x ve y eksenlerımın ısımlerı

plt.title("Poverty Rate Given States")
#buda figurumun baslıgı



```
