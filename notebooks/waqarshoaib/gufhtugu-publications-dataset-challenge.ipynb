```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/gufhtugu-publications-dataset-challenge/GP Orders - 5.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.nunique()
```

```python
df = df.fillna(method='bfill',axis=0)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
sns.set(style="darkgrid")
fig, ax=plt.subplots()
sns.countplot(df['Order Status'],data=df,hue="Order Status")
plt.show()
```

```python
df['City'] = df['City'].str.lower()
```

```python
top_cities = df['City'].value_counts()
top_cities = top_cities.iloc[:10]
top_cities.plot(kind='pie',figsize=(6,6),autopct='%1.1f%%')
```

```python
df2 = df.drop('Book Name', axis=1).join(df['Book Name'].str.split('/', expand=True).stack().reset_index(level=1, drop=True).rename('Book Name'))
```

```python
df2.head(20)
```

```python
top_books = df2['Book Name'].value_counts()
top_books = top_books.iloc[:15].sort_values(ascending=True)
top_books
```

```python
top_books.plot(kind='barh',y='Book Name',figsize=(5,5))
```
