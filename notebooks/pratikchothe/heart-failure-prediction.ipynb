### 5. Data processing 

```python
scaled_df = pd.get_dummies(df,columns=['sex','cp','fbs','restecg','exng','slp','caa','thall']) # this ae all categorical features
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
StandardScaler = StandardScaler()
column_scale = ['age','trtbps','chol','thalachh','oldpeak']
scaled_df[column_scale] = StandardScaler.fit_transform(scaled_df[column_scale])
```

```python
scaled_df.head()
```

```python
X = scaled_df.drop('output', axis = 1)
Y = scaled_df['output']
```

### 6. Train and test data 

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size = 0.3, random_state = 40)
```

### 7. Check K fold cross validatiion 

```python
 def get_score(model, X_train, X_test, Y_train, Y_test):
     model.fit(X_train, Y_train)
     return model.score(X_test, Y_test)
```

```python
# Logistic regression
from sklearn.linear_model import LogisticRegression
get_score (LogisticRegression(max_iter=10000),X_train, X_test, Y_train, Y_test)
```

```python
# Decision Tree
from sklearn.tree import DecisionTreeClassifier
get_score(DecisionTreeClassifier(),X_train, X_test, Y_train, Y_test)
```

```python
# Random Forest
from sklearn.ensemble import RandomForestClassifier
get_score(RandomForestClassifier(),X_train, X_test, Y_train, Y_test)
```

```python
# Support Vector Machine
from sklearn.svm import SVC
get_score(SVC(),X_train, X_test, Y_train, Y_test)
```

### 8. Build Logistic Regression Model 

```python
import sklearn.linear_model as sk
learn = sk.LogisticRegression();          #initializing linear regression model
learn.fit(X_train,Y_train);               #training the linear regression model
Y_predicted = learn.predict(X_test)
score = learn.score(X_test,Y_test);       #testing the linear regression model
print (score)
```

```python
print(classification_report(Y_test, Y_predicted))
```
