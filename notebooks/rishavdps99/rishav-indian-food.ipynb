```python
data=pd.read_csv('/kaggle/input/indian-food-101/indian_food.csv')
data.head()
```

```python
data.describe()
```

```python
data.replace('-1',np.NaN,inplace=True)
item=set()
for i in data['ingredients']:
  item.update(str(i).lower().split(" "))
print(len(item))
```

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

matplotlib.rcParams['font.size']=10
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#000123'

sns.set_style('darkgrid')
state=data.groupby('state').size().to_frame(name="count").reset_index()
sns.barplot(x='count',y='state',data=state)

plt.title('statewise recipe')
plt.xlabel('state')
plt.ylabel('count of recipes')
plt.show()
```

```python
count_diet=data['diet'].value_counts().reset_index()
count_diet['index']
count_diet
```

```python
plt.pie(count_diet.diet,labels=count_diet['index'],radius=1.2,autopct='%0.3f%%',explode=[0,0.4])
plt.title(" Veg Vs Non Veg")
plt.show()
```

```python
count=data['region'].value_counts().reset_index()
count
```

```python
df_regions=data.region.value_counts().reset_index()
sns.barplot(x='region' ,y='index',data=df_regions)
plt.title('Region wise Count of Dishes')
plt.show()
```

```python
df_regions=data.course.value_counts().reset_index()
sns.barplot(x='course' ,y='index',data=df_regions)
plt.title('Count of Cuisines')
plt.show()
```

```python
course_diet=data['course'].value_counts().reset_index()
course_diet['index']
course_diet
```

```python
plt.pie(course_diet.course,labels=course_diet['index'],radius=1.2,autopct='%0.2f%%',explode=[0,0,0,0])
plt.title(" courses")
plt.show()
```

```python
df_cook=(data.prep_time+data.cook_time).to_frame('time').reset_index()
plt.hist(df_cook['time'],np.arange(10,200,10))

plt.title('cook time')
plt.xlabel('Time (minutes)')
plt.ylabel('Recipes')
plt.show()
```
