```python
ps.sqldf("""

SELECT 
d.country,d.release_year,
d.type AS movie_type
FROM df as d
WHERE d.type='Movie'
GROUP BY d.country
LIMIT 10
""")
```

```python
movies=df[df.type.str.contains("Movie")]
movies.head()
```

```python
series=df[df.type.str.contains("TV Show")]
series.head(3)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['figure.figsize']=(8,5)
sns.countplot(df['type'],palette='pink')
plt.title('type of content',fontsize=20)
plt.show()
```

```python
release=movies[movies.release_year>=2015]
sns.countplot(x='release_year',data=release)
plt.title('movies from 2015 ',size='10')
```

**we can see the plots are exceeding the y axis values. So let us set a limit to the y axis**

```python
release=movies[movies.release_year>=2015]
sns.countplot(x='release_year',data=release)
xis=plt.gca()
ymax=movies['type'].value_counts().max()
xis.set_ylim([0,1000])
plt.title('movies from 2015 ',size='13')
```

```python
topdir=movies.groupby(['director'])['show_id'].count().reset_index(name='count').sort_values(by='count',ascending=False)
fig_dims=(18,7)
fig, ax = plt.subplots(figsize=fig_dims)

sns.pointplot(x='director',y='count',data=topdir.head(10))
```

# **Best Rated Movies**

```python
toprate=movies.groupby(['rating'])['show_id'].count().reset_index(name='count').sort_values(by='count',ascending=False)
fig_dims=(18,7)
fig, ax = plt.subplots(figsize=fig_dims)

sns.pointplot(x='rating',y='count',data=toprate.head(10))
```

**Tv-MA has the highest rating**
