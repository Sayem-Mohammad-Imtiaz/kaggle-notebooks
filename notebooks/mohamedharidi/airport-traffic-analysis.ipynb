# Data Cleaning
* replace the capital letters with small letters and the space with dash (_)
* change date type from str to datetime 

```python
df.rename(columns=lambda x:x.strip().lower(),inplace=True)
```

```python
df.date=pd.to_datetime(df.date)
```

```python
days={0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thursday',4:"Friday",5:'Saturday',6:'Sunday'}
```

```python
df['weekday']=df.date.apply(lambda x:days.get(x.weekday()))
# make a new coulmn
```

```python
df["lon"]=df.centroid.apply(lambda x:x.split()[0].replace('POINT(',' '))
df["lat"]=df.centroid.apply(lambda x:x.split()[1].replace(')',' '))
```

```python
fig_map = go.Figure(data=[
    go.Scattergeo(lat=df['lat'],
             lon=df['lon'],
             mode='markers',
             hoverinfo='text',
             marker=dict(color="darkred", size=5, opacity=0.9))
])
fig_map.update_geos(
    landcolor='white',
    showocean=True,
    oceancolor='white',
    lakecolor='blue',
)
fig_map.update_layout(
    width=800,
    height=400,
    template="plotly_white",
    margin=dict(r=0, t=0, b=0, l=0),
    annotations=[
        dict(
            text="",
            showarrow=False,
            xref="paper",
            yref="paper",
            x=0,
            y=0)
    ])

fig_map.show()
```
