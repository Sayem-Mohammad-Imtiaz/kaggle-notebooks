# Data Cleaning
* replace the capital letters with small letters and the space with dash (_)
* drop_duplicates
* change member_number type from int to str 
* split a date to year and month and day



```python
df.rename(columns=lambda x:x.strip().lower(),inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["member_number"]=df.member_number.astype(str)
```

```python
df["date"]=df.date.astype('datetime64')
```

```python
df['year']=df.date.apply(lambda x:x.year)
df['month']=df.date.apply(lambda x:x.month)
days={0:'Monday',1:'Tuesday',2:'Wednesday',3:'Thursday',4:'Friday',5:'Saturday',6:'Sunday'}
df['day']=df.date.apply(lambda x:days.get(x.weekday()))
```

# Exploratory the  Data 

```python
df.query('year==2015').itemdescription.value_counts().head(15)
```

```python
fig = px.sunburst(df, path=["year",'month', 'day'])
fig.show()
```

there is no big difference between sales in each day 
with more sales in 2015


```python
fig=px.histogram(df, x="itemdescription",animation_frame='month')
fig.update_xaxes(range=(-.5,14.5))
fig.update_traces(marker_color='rgb(158,202,225)', marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5, opacity=0.6)
fig.update_layout(title='top 15 item per month',barmode='stack', xaxis={'categoryorder':'total descending'},)

fig.show()
```

```python
fig=px.histogram(df, x="itemdescription",animation_frame='month',facet_col='day')
fig.update_xaxes(range=(-.5,14.5))
fig.update_traces(marker_color='rgb(158,202,225)', marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5, opacity=0.6)
fig.update_layout(title='top 15 item per month  each day',barmode='stack', xaxis={'categoryorder':'total descending'},)

fig.show()
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
app = dash.Dash(__name__)
days = df.day.unique()

app.layout = html.Div([
    dcc.Dropdown(
        id="dropdown",
        options=[{"label": x, "value": x} for x in days],
        value=days[0],
        clearable=False,
    ),
    dcc.Graph(id="bar-chart"),
])

@app.callback(
    Output("bar-chart", "figure"), 
    [Input("dropdown", "value")])
def update_bar_chart(day):
    mask = df["day"] == day
    fig=px.histogram(df, x="itemdescription",animation_frame='month')
    fig.update_xaxes(range=(-.5,14.5))
    fig.update_layout(title='top 15 item per month',barmode='stack', xaxis={'categoryorder':'total descending'},)

    
    return fig
app.run_server(debug=True)
```
