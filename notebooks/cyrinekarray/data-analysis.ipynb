**Function that enumerate scores with parametre data and hue**

```python
def function_plot_all_scores(title, data, hue):
    fig, axes = plt.subplots(1, 3, figsize=(24,6))
    fig.suptitle(title)

    scores = ['math score', 'reading score', 'writing score']
    for i, score in enumerate(scores):
        sns.kdeplot(ax=axes[i], x=score, data=data, hue=hue, linewidth=2, fill=True, alpha=.2)
        axes[i].set_title(score)  
        plt.setp(axes[i].get_legend().get_texts(), fontsize='11')
        plt.setp(axes[i].get_legend().get_title(), fontsize='13')
```

```python
function_plot_all_scores('Scores for test preparation', df, 'test_preparation')
```

```python
function_plot_all_scores('Scores for lunch', df, 'lunch')
```

```python
function_plot_all_scores('Scores for education level', df, 'education_level')
```

**Convert categorical data into dummy variables**

```python
df = pd.get_dummies(df, columns=['gender','race_ethnicity', 'education_level', 'lunch', 'test_preparation'],
              drop_first=True)
```

```python
df.head()
```

**Combine score of student in total score**

```python
df['score'] =  df[['math score', 'reading score', 'writing score']].apply(np.mean, axis=1)
df['score'] = df['score'].apply(lambda x : int(x))
df.drop(['math score', 'reading score', 'writing score'], inplace=True, axis=1)
```

```python
df.head()
```

```python
X = df.drop(['score'], axis=1)
Y = df['score']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4, random_state=0)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
train_data = lgb.Dataset(X_train, label=y_train)
valid_data = lgb.Dataset(X_test, label=y_test)
parameters = {
    'objective': 'regression',
    'metric': 'mse',#mean squared error
    'is_unbalance': 'true',
    'boosting': 'gbdt',
    'num_leaves': 63,
    'feature_fraction': 0.5,
    'bagging_fraction': 0.5,
    'bagging_freq': 20,
    'learning_rate': 0.01,
    'verbose': -1
}
model_lgbm = lgb.train(parameters,
                           train_data,
                           valid_sets=valid_data,
                           num_boost_round=1000,
                           early_stopping_rounds=50)
```

```python
lgb.plot_importance(model_lgbm, figsize=(26, 12))
```

```python
predictions = model_lgbm.predict(X_test)
print('Mean squared error:', mean_squared_error(y_test, predictions))
print('Squared Mean squared error:', np.sqrt(mean_squared_error(y_test, predictions)))
print('Mean absolute error:', mean_absolute_error(y_test, predictions))
```
