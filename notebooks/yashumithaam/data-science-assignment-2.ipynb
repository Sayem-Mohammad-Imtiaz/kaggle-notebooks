# Data Science Assignment - Data Exploration
Here, we will be exploring data from a dataset and drawing conclusions from that data related to the hypothesis we create.The dataset used here is the Housing Sales in King County dataset which contains many parameters of the house (No of bedrooms,square feet,no of bathrooms etc) with the selling price of the house.

**Hypothesis : The selling price of a house is dependent upon the number of bathrooms it has.**<br>
This statement has been made from general observation and logic.<br>
Let us visualize the above data and validate our Hypothesis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Let us import the necessary libraries.<br>Pandas is used to read the dataset which is in CSV format. Numpy is required for representation and array operations after reading the data. Matplotlib is used to plot and visualize the data. Now let's import the dataset.

```python
data  = pd.read_csv("/kaggle/input/housesalesprediction/kc_house_data.csv")
```

Now the dataset is imported and stored as numpy array. Let us perform a cleaning operation on dataset using dropna(). This removes the rows where NULL or empty values are present. 

```python
data.dropna()
```

The dataset has now been cleaned and is fit for evaluation. The next step is to extract parameters from the dataset which we should explore. According to our hypothesis we require the parameters Number of bedrooms, Number of Bathrooms, the Condition of the house and Selling Price of the house.

```python
brooms = data[ 'bathrooms'] #extract the values of columns with the name 'bathrooms'
price = data['price'] #extract the values of columns with the name 'SalePrice'
```

Once the require data is extracted we will have to visualize the data to make a conclusion and validate our hypothesis. For this we use Matplotlib to plot the data along the axes and show the graph.

```python
plt.plot(brooms,price)
plt.show()
```

The graph above is highly irregular with respect to the relation between the two variables.Now let us try to understand the graph and use it to validate our hypothesis.
On observing the graph,
Consider a house with one bathroom : We can observe that the same house is available for many difference prices. In fact the same one bathroom is present in cheapest as well as the most costliest.
Similarly, the same results apply for the other values along x-axis as well.
This shows us the conclusion that the selling price of a house is not entirely dependent on the number of bathrooms but it also depends on other factors that are present in the dataset like sqft_living,number of floors, condition of the house, year built etc.<br>
**Hence we can conclude that the hypothesis we generated is partially true.**
