```python
# read data and show titles 
data = pd.read_csv("../input/bitcoin-historical-data/bitstampUSD_1-min_data_2012-01-01_to_2020-12-31.csv")
data.head()
```

```python
# I leart the information of data (column, dtype, memory usage...)
data.info()
```

```python
# to look at the corrolation between the headings
data.corr()
```

```python
# correlation map
f, ax = plt.subplots(figsize=(13,13))
sns.heatmap(data.corr(), annot = True, linewidths = 1, fmt=".1f", vmax=1, vmin=0, cmap='coolwarm', ax=ax)
plt.show()
```

```python
data.columns
```

```python
# Line Plot 
data.Open.plot(kind = "line", color= "r", label = "Open", linewidth=1, alpha = 0.5, grid= True, linestyle= ":")
data.High.plot(color= "g", label = "High", linewidth=1, alpha = 0.5, grid= True, linestyle= "-.")
plt.legend(loc="upper right")
plt.xlabel("Open")
plt.ylabel("High")
plt.title("Line Plot")
plt.show()
```

```python
#Scatter
data.plot(kind="scatter", x="Open", y="High", alpha=0.3, color= "blue")
plt.xlabel("Open")
plt.ylabel("High")
plt.title("Scatter Plot")
plt.show()
```

```python
#Scatter
data.plot(kind="scatter", x="Timestamp", y="Weighted_Price", alpha=0.3, color= "green")
plt.xlabel("Timestamp")
plt.ylabel("Weighted_Price")
plt.title("Scatter Plot")
plt.show()
```

```python
# Histogram
data.Open.plot(kind="hist", bins= 100, figsize=(12,12))
plt.title("Histgram Plot")
plt.show()
```
