```python
# Peep at the competitions file header
df_Competitions = pd.read_csv('../input/meta-kaggle/Competitions.csv')
df_selectedCompetitions = df_Competitions.sort_values(by=['TotalTeams'], ascending=False).head(20)
df_selectedCompetitions.to_csv('selected_Competitions4.csv',index=False)
df_selectedCompetitions.info()
```

```python
selected_Competitions = df_selectedCompetitions['Id'].values.tolist()
selected_Competitions = list(set(selected_Competitions))
selected_Competitions
```

```python
df_Teams = pd.read_csv('../input/meta-kaggle/Teams.csv')

df_SelectedTeams = df_Teams[df_Teams.CompetitionId.isin(selected_Competitions)]
df_SelectedTeams.to_csv('selected_Teams4.csv',index=False)
df_SelectedTeams.info()
```

```python
teamLeaders =  df_SelectedTeams['TeamLeaderId'].fillna(0).astype(int).values.tolist()
teamLeaders = list(set(teamLeaders))
SelectedTeams = df_SelectedTeams['Id'].values.tolist()
SelectedTeams
```

```python
df_TeamMemberships = pd.read_csv('../input/meta-kaggle/TeamMemberships.csv')
df_selectedMemberships = df_TeamMemberships[df_TeamMemberships.TeamId.isin(SelectedTeams)]

df_selectedMemberships.to_csv('selected_Memberships4.csv',index=False)
df_selectedMemberships.info()
```

```python
selectedMembers = df_selectedMemberships['UserId'].values.tolist()

selectedCompetitors = selectedMembers + teamLeaders
selectedCompetitors = list(set(selectedCompetitors))
df_selectedCompetitors = df_users[df_users.Id.isin(selectedCompetitors)]
df_selectedCompetitors.to_csv('selected_Users4.csv',index=False)
df_selectedCompetitors.head()
```
