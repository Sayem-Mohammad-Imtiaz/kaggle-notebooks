```python
# 0: Load Data
wine = pd.read_csv('/kaggle/input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
```

```python
# 1: Preview 5 Data
wine.head(5)
```

```python
# 2: Number of wines and Attributes
print('Number of wines: ' + str(wine.shape[0]))
print('Number of wine attributes: ' + str(wine.shape[1]))
```

```python
# 3: Top 5 Highest alcohol content
print('Top 5 alcohol content: ') 
wine.sort_values('alcohol', ascending = False).iloc[:5]
```

```python
# 4: Bottom 5 Lowest alcohol content
print('Bottom 5 alcohol content: ')
wine.sort_values('alcohol').iloc[:5]
```

```python
# 5: Wine with highest clorides (highest preservatives)
print("Wine with highest chlorides: ")
wine[ wine.chlorides == wine.chlorides.max() ]
```

```python
# 6: Average alcohol contents
print("Average wine alcohol content accross " + str(wine.shape[0]) + " wines is " + str(round(wine.alcohol.mean())))
```

```python
# 7: Visual representation of # of wines by quality
wine.groupby('quality') \
    .size() \
    .plot.pie()
```
