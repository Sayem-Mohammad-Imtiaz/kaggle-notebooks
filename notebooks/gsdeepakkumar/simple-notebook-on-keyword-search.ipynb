In this kernel, I am interested to see how many of them have actually mentioned the vaccine name in their comments. I am using re module to create a basic set of keywords and I loop the reddit comments to flag the keywords.It is seen that only a very few people have mentioned the vaccine name and the most talked about vaccine are Pfizer and Moderna.
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
import re
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('../input/reddit-vaccine-myths/reddit_vm.csv')
```

Remove rows with no comments in body.

```python
data_clean=data.loc[~(data['body'].isna())]
```

Lets focus our analysis on comments which have vaccine names in them - Pfizer,Moderna,AstraZeneca,Sputnik V,CoronaVac,Covaxin.


```python
vaccines=['pfizer','moderna','astra','sputnik','coronavac','covaxin','sinovac']
print(re.compile(r'\b(?:%s)\b' % '|'.join(vaccines)))
```

```python
def search(text):
    return set([i for i in text.lower().split() if re.search(re.compile(r'\b(?:%s)\b' % '|'.join(vaccines)),i)])
```

```python
data_clean.loc[~(data_clean['body'].isna()),'kwd']=data_clean.loc[~(data_clean['body'].isna()),'body'].apply(lambda x:search(x))
```

```python
data_clean['kwd']=data_clean['kwd'].apply(lambda x:np.nan if len(x)==0 else x)
```
