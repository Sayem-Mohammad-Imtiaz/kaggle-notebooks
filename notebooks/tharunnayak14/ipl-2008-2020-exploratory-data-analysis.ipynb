# City

```python
matches['city'].value_counts().iplot()
```

```python
sum(matches['city'].isnull())
```

```python
matches[matches['city'].isnull()]
```

```python
def replace_city(cols):
    city = cols[0]
    venue = cols[1]
    if pd.isnull(city):
        return venue.split()[0]
    else:
        return city   
```

```python
matches['city'] = matches[['city','venue']].apply(replace_city,axis=1)
```

```python
sum(matches['city'].isnull())
```

# Result margin

```python
matches['result_margin'].iplot(kind='hist')
```

```python
sns.displot(matches['result_margin'],bins=30,kde=True, height=8,aspect=1.8)
```

```python
matches.groupby('winner').mean()['result_margin'].iplot()
```

```python
df = matches.groupby('winner').mean()['result_margin']
d = {}
for i in df.index:
    d[i] = df.loc[i]

print(d)
```

```python
sum(matches['result_margin'].isnull())
```

```python
matches[matches['result_margin'].isnull()]
```

```python
def impute_result_margin(cols):
    result_margin = cols[0]
    winner = cols[1]
    
    if pd.isnull(result_margin) and not pd.isnull(winner):
        return d[winner]
    else:
        return result_margin
    
```

```python
matches['result_margin'] = matches[['result_margin','winner']].apply(impute_result_margin,axis=1)
```

```python
sum(matches['result_margin'].isnull())
```
