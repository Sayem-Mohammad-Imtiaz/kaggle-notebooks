`ax.flatten()` converts 2d ax array into 1d ax array   
However graph looks frustrating   
you can solve this porblem by `plt.tight_layout()`

```python
fig, ax = plt.subplots(2, 3, figsize=(8, 5))
for i,iax in np.ndenumerate(ax.flatten()):
    iax.set_title(f'this {i[0]}')
plt.tight_layout()
plt.show()
```

perfect!!   
if you want custom size subplot(not same size),  
`plt.subplot2grid()` can do this 

```python
fig = plt.figure(figsize=(8, 5)) # initialize figure

ax = [None for _ in range(6)] # list to save many ax for setting parameter in each

total_size=(3,4)

ax[0] = plt.subplot2grid(total_size, (0,0), colspan=4)
ax[1] = plt.subplot2grid(total_size, (1,0), colspan=1)
ax[2] = plt.subplot2grid(total_size, (1,1), colspan=1)
ax[3] = plt.subplot2grid(total_size, (1,2), colspan=1)
ax[4] = plt.subplot2grid(total_size, (1,3), colspan=1,rowspan=2)
ax[5] = plt.subplot2grid(total_size, (2,0), colspan=3)


for i,iax in np.ndenumerate(ax): 
    iax.set_title(f'ax[{i[0]}]') # make ax title for distinguish:)
    iax.set_xticks([]) # to remove x ticks
    iax.set_yticks([]) # to remove y ticks
    
fig.tight_layout()
plt.show()
```

if you want to more custom size,you can programmingly use `plt.add_axes()`

```python
fig = plt.figure(figsize=(8, 5))

ax = [None for _ in range(3)]


ax[0] = fig.add_axes([0.1,0.1,0.8,0.4]) # x, y, dx, dy
ax[1] = fig.add_axes([0.15,0.6,0.25,0.6])
ax[2] = fig.add_axes([0.5,0.6,0.4,0.3])

for ix in range(3):
    ax[ix].set_title('ax[{}]'.format(ix))
    ax[ix].set_xticks([])
    ax[ix].set_yticks([])

plt.show()
```

# 2. Colormap

* diverging
* qualitative
* sequential

```python
gradient = np.linspace(0, 1, 256)
gradient = np.vstack((gradient, gradient))


def plot_color_gradients(cmap_category, cmap_list):
    # Create figure and adjust figure height to number of colormaps
    nrows = len(cmap_list)
    figh = 0.35 + 0.15 + (nrows + (nrows-1)*0.1)*0.22
    fig, axes = plt.subplots(nrows=nrows, figsize=(6.4, figh))
    fig.subplots_adjust(top=1-.35/figh, bottom=.15/figh, left=0.2, right=0.99)

    axes[0].set_title(cmap_category + ' colormaps', fontsize=14)

    for ax, name in zip(axes, cmap_list):
        ax.imshow(gradient, aspect='auto', cmap=plt.get_cmap(name))
        ax.text(-.01, .5, name, va='center', ha='right', fontsize=10,
                transform=ax.transAxes)

    # Turn off *all* ticks & spines, not just the ones with colormaps.
    for ax in axes:
        ax.set_axis_off()
```

**perceptually uniform sequential**   
it is used often for signal data

```python
pus=('Perceptually Uniform Sequential', [
            'viridis', 'plasma', 'inferno', 'magma', 'cividis'])
plot_color_gradients(pus[0], pus[1])
```

```python
sq=('Sequential', [
            'Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
            'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
            'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn'])
plot_color_gradients(sq[0], sq[1])
```
