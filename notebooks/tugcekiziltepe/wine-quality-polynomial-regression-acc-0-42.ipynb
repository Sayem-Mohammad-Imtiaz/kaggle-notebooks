# 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<a id = '2'> </a> <br>
# 2. Load the Data

```python
df = pd.read_csv('/kaggle/input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
df
```

```python
df.info()
```

All data is numeric. That's why there is no need to encode.

<a id = '3'> </a> <br>
# 3. Taking Care of Missing Data
There are no missing values in this data as you can see below.

```python
df.notnull().all()
```

## Set Independent and Dependent Variable

```python
#Independent Variable
x = df.iloc[:, :-1].values
#Dependent Variable
y = df.iloc[:, -1].values
```

<a id = '4' > </a> <br>
# 4. Split the Dataset into Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 19)
```

<a id = '5'> </a> <br>
# 5. Training the Polynomial Linear Regression Model on the Training Set

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
x_poly = poly_reg.fit_transform(x_train)
linear_reg = LinearRegression()
linear_reg.fit(x_poly, y_train)
y_pred = linear_reg.predict(poly_reg.transform(x_test))
```

<a id = '6'> </a> <br>
# 6. Predicting the Test Set Results

```python
y_pred = linear_reg.predict(poly_reg.transform(x_test))
```

```python
y_pred
```

<a id = '7'> </a> <br>
# 7. Calculate the Accuracy

Here I will use R squared for calculating the accuracy.

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
