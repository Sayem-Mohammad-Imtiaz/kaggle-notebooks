# 1. Import Libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
```

<a id = '2'> </a> <br>
# 2. Load the Data

```python
df = pd.read_csv('/kaggle/input/car-price-prediction/CarPrice_Assignment.csv')
```

```python
df
```

```python
df.info()
```

Object types are strings. I am going to use One Hot Encoding for them.   
For numeric variable, there is no need to encode.

<a id = '3'> </a> <br>
# 3. Drop Unnecessary Columns

Car ID, symboling, car name does not affect the price.


```python
df.drop(['car_ID', 'symboling', 'CarName'], axis=1, inplace=True)
x = df.iloc[:, :-1].values #independent variables
y = df.iloc[:, -1].values #dependent variable, price.
```

```python
x
```

```python
y
```

<a id = '4'> </a> <br>
# 4. Take Care of Missing Data
As you can see below, there is no missing(null) value in this data.

```python
df.notnull().all()
```

<a id = '5'> </a> <br>
# 5. Encoding Categorical Data
One Hot Encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer( transformers= [('encoder', OneHotEncoder(), [0, 1 ,2 ,3 ,4, 5, 11, 12, 14])], remainder = 'passthrough')
x = np.array(ct.fit_transform(x))
```

```python
x
```

<a id = '6'> </a> <br>
# 6. Split the Dataset into Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 15)
```

<a id = '7'> </a> <br>
# 7. Training the Multiple Linear Regression Model on the Training Set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

<a id = '8'> </a> <br>
# 8. Predicting the Test Set Results

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision = 2) # 2 numbers after comma
```

```python
y_pred
```

<a id = '9'> </a> <br>
# 9. Calculate the Accuracy

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

### As a result, the accuracy is %85.
