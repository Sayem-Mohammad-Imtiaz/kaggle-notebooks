```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/c/titanic/train.csv')
df.head()
```

```python
# Visualizing Missing Data
import missingno as msno

missing_data = msno.bar(df, figsize=(6,4))
print(df.info())
print('------'* 20 , '\n\n')
print(df.isnull().sum())
print('------'* 20 , '\n\n')
print(missing_data)
```

#### Handling Missing Data

```python
## Drop All the missing Record Values

df1 = df.dropna()
print('origional shape: ',df.shape, '----->', 'New Shape',df1.shape)
'''
You can see the the difference between the origional and new shape of the data
891 to 183 records...
We are discarding most of the data and hence our model will not have sufficient data,
to learn on. Hence the predictions will be poor.
'''
```

```python
## Droping Features
msno.bar(df, figsize=(6,4))
print('Majority of the values in cabin feature are missing So we will drop the feature')

df2 = df.drop('Cabin', axis=1)
```

```python
print(df2.isnull().sum())
```

```python
#### Handling missing values with the mean Age from each passanger class.

plt.figure(figsize=(8,7))
sns.boxplot(x = df2['Pclass'], y = df2['Age'])
```

if you look at the age distribution for each passanger class you'll see that 
average number of people in class 3 are young compared to class 2 and 3

```python
# To check the median age for each class
c1 = df2[df['Pclass'] == 1]
c2 = df2[df['Pclass'] == 2]
c3 = df2[df['Pclass'] == 3]
print(f'Class 1 average age is {int(c1.Age.median())}, Class 2 average age is {int(c2.Age.median())} and Class 3 average age is {int(c3.Age.median())} ')
```

```python
df2['Age'] = df2['Age'].fillna(df2.groupby('Pclass')['Age'].transform('median'))
df2.isnull().sum()
```

```python
# I will discard the two remaining values for embarked

df2.dropna(inplace=True)
df2.isnull().sum()

# Now we have handled all the missing values....
```
