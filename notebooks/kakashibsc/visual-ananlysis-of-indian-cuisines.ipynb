# Basic Visualisation of Indian Cuisines.

# Importing basic plotting libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

# Creating a DataFrame

```python
ind_food = pd.read_csv('/kaggle/input/indian-food-101/indian_food.csv')
```

The head of the DataFrame - 'ind_food'

```python
ind_food.head()
```

Checking if any null values are present

```python
ind_food.isnull().sum()
```

Dropping null values

```python
ind_food.dropna(inplace = True)
```

```python
ind_food.isnull().sum()
```

# Visualisation

Catergories with respect to taste of Indain Cuisines

```python
sns.countplot(ind_food['flavor_profile'])
```

Removal of '-1' category in the 'flavour_profile'

```python
ind_food.drop(ind_food[ind_food['flavor_profile']=='-1'].index, inplace = True)
sns.countplot(ind_food['flavor_profile'])
```

As per the graph,
* Indians tend to eat more of spicy food camparitively with other countries.
* Sweets are the next best cuisine avaialble with a great demand in India especially during festive seasons.

A plot on COURSES of food Indians prefer.

```python
sns.countplot(ind_food['course'])
```

As per the graph,
* Main course is observed to be dominant. Main course may include rice items, chapathis, etc.
* Dessert come next in the line, as Indians love sweets or something soothing after a heavy meal or at celebrations.
* Starters aren't that preferred compared to snacks.

Plot of count of Indian Cuisines available in accordance with Regions of India.

```python
ind_food.drop(ind_food[ind_food['region']=='-1'].index, inplace = True)
sns.countplot(ind_food['region'])
```

* West region of India have a huge number of different type of Indian Cuisines that add up to the famous Indian Dishes prepared world-wide.
* South Region of India give in the send large count of many different Indian Cuisines available, rice being the dominant and staple food of South India.
* Central region of the country has the least count of available Indian Cuisines.

A pie plot of Vegetarian and Non-Vegetarian categories in India according to the dataset.

```python
ind_food['diet'].value_counts().plot(kind = 'pie', figsize = (5,5))
```

* More vegetarians are present in India compared to the Non-Vegeratians

A plot of cook-time taken to prepare two categories of Indian Cuisines w.r.t 'diet'.

```python
sns.barplot(x = 'diet',y = 'cook_time', data = ind_food)
```

* Vegetarian dishes tend to take more time to cook compared to Non-Vegetarian dishes
* More than 30 minutes of cook-time is required for delicious veg dish.

A plot on cook time w.r.t categories of 'courses'

```python
sns.barplot(x = 'course',y = 'cook_time', data = ind_food)
```

* The cook time for dessert is more compared to all the food courses.

A plot on preperation time w.r.t categories of 'courses'.

```python
sns.barplot(x = 'course',y = 'prep_time', data = ind_food)
```

* The preperation time for starters is more compared to other food courses.
THe preperation becomes a mundane task and thus starters are not preferred to be cooked often compared to other course.

A plot of preperation time taken to prepare two categories of Indian Cuisines w.r.t 'diet'.

```python
sns.barplot(x = 'diet',y = 'prep_time', data = ind_food)
```

* The preperation time for vegetarian and non-vegetarian courses are approx similar.

# Creating new DataFrames for the following categories of flavour profiles:
# * Sweet
# * Spicy
# * Bitter

```python
sweets = ind_food[(ind_food['flavor_profile']=='sweet')]
spicy_ = ind_food[(ind_food['flavor_profile']=='spicy')]
bitter_ = ind_food[(ind_food['flavor_profile']=='bitter')]
sour_ = ind_food[(ind_food['flavor_profile']=='sour')]
```

```python
plt.figure(figsize=(20,5))
plt.xticks(rotation=75)
sns.countplot(sweets['state'])
```

* By the above plot, we can observe that West Bengal consumes more sweets compared to other states.

2) A plot to see which state consumes the more spicy foods.

```python
plt.figure(figsize=(20,5))
plt.xticks(rotation=75)
sns.countplot(spicy_['state'])
```

* As per above plot, Punjab consumes more spicy food compared to other states

3) A plot to see which state consumes the more bitter flavoured foods.

```python
plt.figure(figsize=(20,5))
plt.xticks(rotation=75)
sns.countplot(bitter_['state'])
```

* As per above plot, only two states are observed to consume bitter flavour profiled dish

4) A plot to see which state consumes the more sour profiled foods.

```python
plt.figure(figsize=(20,5))
sns.countplot(sour_['state'])
```

* As per above plot, only Gujarat state is observed to cosume sour profiled foods.

# Categorising different states according to regions of the country.

```python
east = ind_food[(ind_food['region']=='East')]
west = ind_food[(ind_food['region']=='West')]
north = ind_food[(ind_food['region']=='North')]
south = ind_food[(ind_food['region']=='South')]
north_east = ind_food[(ind_food['region']=='North East')]
central = ind_food[(ind_food['region']=='Central')]
```

1) The states under the Northern region of the country

```python
north = north.reset_index()
ingredients = []
for i in range(0,len(north)):
    text = north['state'][i].split(',')
    text = ','.join(text)
    ingredients.append(text)
    text = ' '.join(ingredients)

wordcloud = WordCloud(width = 400, height = 400, colormap = 'seismic'
                      ,background_color ='white', 
                min_font_size = 8).generate(text)                  
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis('off') 
plt.show()
```

Ingredients used by the Northern region of the country

```python
ingredients = []
for i in range(0,len(north)):
    text = north['ingredients'][i].split(',')
    text = ','.join(text)
    ingredients.append(text)
    text = ' '.join(ingredients)

wordcloud = WordCloud(width = 400, height = 400, colormap = 'seismic'
                      ,background_color ='white', 
                min_font_size = 8).generate(text)                  
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis('off') 
plt.show()
```
