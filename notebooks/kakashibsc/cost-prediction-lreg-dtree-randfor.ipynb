# Building the predictive model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, r2_score
```

```python
ins_table.head()
```

Sincce sex is in strings format for 'male' and 'female', lets change it to numerical representation.
* Let 0 represent female
* Let 1 represent male

```python
ins_table['sex'] = pd.get_dummies(ins_table['sex'], drop_first=True)
```

```python
ins_table.head()
```

Similarly to get numerical represntation of columns for 'smoker' and 'region' are as follow:
* For 'smoker' column:
    1. Value 1 for 'yes'
    2. Value 0 for 'no'
* For 'region' column:
    1. Value 1 for 'northeast'
    2. Value 2 for 'northwest'
    3. Value 3 for 'southeast'
    4. Value 4 for 'southwest'

```python
ins_table['smoker'] = pd.get_dummies(ins_table['smoker'], drop_first=True)
```

```python
def reg(arr):
    if arr['region'] == 'northeast':
        return 1
    elif arr['region'] == 'northwest':
        return 2
    elif arr['region'] == 'southeast':
        return 3
    else:
        return 4
ins_table['region'] = ins_table.apply(reg, axis = 1)
```

```python
ins_table.head()
```

```python
X = ins_table.drop('charges',axis = 1)
y = ins_table['charges']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X ,y, test_size = 0.3, random_state = 100)
```

# 1. Linear Regression

```python
lin = LinearRegression()
```

```python
lin.fit(X_train, y_train)
```

```python
pred_1 = lin.predict(X_test)
```

```python
print(r2_score(y_test, pred_1))
```

# 2. Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
dtree = DecisionTreeRegressor()
dtree.fit(X_train, y_train)
```

```python
pred_2 = dtree.predict(X_test)
```

```python
print(r2_score(y_test, pred_2))
```

Accuracy 70.4%

# 3. Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
rfor = RandomForestRegressor()
rfor.fit(X_train, y_train)
```

```python
pred_3 = rfor.predict(X_test)
```

```python
print(r2_score(y_test, pred_3))
```

Accuracy 84.68%

# Among the three basic regressors Random Forest Regressor gave better accuracy and works well on this dataset
