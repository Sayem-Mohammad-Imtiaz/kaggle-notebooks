```python
#Read and manage data

df = pd.read_csv('/kaggle/input/diabetes/diabetes.csv')
test_data = df.head(700)
test_data.insert(0, 'ones',1)
cols = test_data.shape[1]
x = test_data.iloc[:,0:cols-1]
y = test_data.iloc[:,cols-1:cols]

x = np.array(x.values)
y = np.array(y.values)
theta = np.zeros(9)
```

```python
###################Functions for Logistic Regression#############################
#Segmoid
def sigmoid(z):
    h = 1/(1 + np.exp(-z))
    return h


#cost func to calculate the error rate
def cost(theta, X,Y):
    theta = np.matrix(theta)
    X = np.matrix(X)
    Y = np.matrix(Y)
    first = np.multiply(-Y, np.log(sigmoid(X * theta.T)))
    second = np.multiply((1 - Y), np.log(1 - sigmoid(X * theta.T)))
    return np.sum(first - second) / (len(X))


#gradiant for optimizing the error rate
def gradient(theta, X, Y):
    theta = np.matrix(theta)
    X = np.matrix(X)
    Y = np.matrix(Y)
    
    parameters = int(theta.ravel().shape[1])
    grad = np.zeros(parameters)
    
    error = sigmoid(X * theta.T) - Y
    
    for i in range(parameters):
        term = np.multiply(error, X[:,i])
        grad[i] = np.sum(term) / len(X)
    
    return grad

################################
#Optimize the cost by gradient
result = opt.fmin_tnc(func=cost, x0=theta, fprime=gradient, args=(x, y),disp=False)
```

```python
#function to predict and test my model
def predict(Theta, X):
    per = []
    prob = sigmoid(np.dot(X, Theta.T))
    #print ('prob',prob)
    for i  in prob:
        if i >= 0.5:
            per.append(1)
        else:
            per.append(0) 
    return per
```

```python
predictions = predict(result[0],x)
prediction = np.matrix(predictions)

#For loop its  just for fun to calc my acurracy
sum = 0
for i in range(700):
    if predictions[i] != y[i,0]:
        sum += 1
    else:
        sum = sum
        
perc = 100 - ((sum/700)*100)
print ('accuracy = {0}%'.format(perc))
```
