Data Processing

After exploring the dataset, use dummy varieble to covert data into categorical form.

```python
data.columns
```

```python
dataset = pd.get_dummies(data,columns = ['sex','cp','fbs','restecg','exang','slope','ca','thal'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[columns_to_scale] = standardScaler.fit_transform(dataset[columns_to_scale])
```

```python
dataset.head()
```

```python
y = dataset['target']
X = dataset.drop(['target'], axis = 1)
```

```python
from sklearn.model_selection import cross_val_score
knn_scores = []
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score=cross_val_score(knn_classifier,X,y,cv=10)
    knn_scores.append(score.mean())
```

```python
plt.plot([k for k in range(1, 21)], knn_scores, color = 'red')
for i in range(1,21):
    plt.text(i, knn_scores[i-1], (i, knn_scores[i-1]))
plt.xticks([i for i in range(1, 21)])
plt.xlabel('Number of Neighbors (K)')
plt.ylabel('Scores')
plt.title('K Neighbors Classifier scores for different K values')
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 12)
score=cross_val_score(knn_classifier,X,y,cv=10)
```

```python
score.mean()
```

Random forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randomforest_classifier= RandomForestClassifier(n_estimators=10)

score=cross_val_score(randomforest_classifier,X,y,cv=10)
```

```python
total_cols = len(data.columns)
total_cols
x = data.values[:,:total_cols-1]
y = data.values[:,total_cols-1]
```

```python
#split the data into training set and testing set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state = 0)
print(x_train)
```

```python
# import the regressor 
from sklearn.tree import DecisionTreeRegressor
```

```python
# fit the regressor with X and Y data
regressor.fit(x_train, y_train) 
```
