# IMPORTING LIBRARIES:

```python
import warnings
warnings.filterwarnings('ignore')

# basic libraries
import numpy as np
import pandas as pd

#Visualization Libraries
import matplotlib
import missingno as msno
import matplotlib.pyplot as plt
import seaborn as sn
import plotly.graph_objs as go
```

```python
df = pd.read_csv("../input/pima-indians-diabetes-database/diabetes.csv")
print(f'we have {df.shape[1]} instances and the following {df.shape[0]} attributes')
df.head()
```

```python
df.isnull().sum()
```

# DESCRIPTIVE STATISTICS

```python
df.describe().T.style.set_table_attributes("style='display:inline'").bar(color='blue')
```

### Replace the zeros with nan values to get the null accuracy

```python
df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']] = df[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']].replace(0,np.nan)
df.head()
```

```python
color = ['red','blue']*4
plt.title('Visualization of Null Values in Dataset',fontsize=28)
msno.bar(df,sort='descending',filter='top',color=color)
```

### Null accuracy

```python
null_accuracy = max(df.Outcome.values.mean(),1-df.Outcome.values.mean())
print(f'Null Accuracy: {null_accuracy * 100}')
print('\nThis is the Baseline accuracy our model we need to cross')
```
