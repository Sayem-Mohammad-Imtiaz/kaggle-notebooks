# Visualization 


```python
countries = df['country'].value_counts()[df['country'].value_counts(normalize=True)> 0.005]
```

```python
plt.figure(figsize=(20,10))
plt.title('Amount of content by country', fontsize=18)
plt.tick_params(labelsize=14)
sns.barplot(y=countries.index, x=countries.values)
plt.show()
```

```python

px.pie(countries.sort_values(ascending=False).iloc[:10],values=countries.values,names=countries.index,title='Content quantity by country in percentage')
```

above plot can be made more efficient with this code

```python
TVshows= df[df['type'] == 'TV Show']
Movies= df[df['type'] == 'Movie']
```

```python
TVshows_progress = TVshows['release_year'].value_counts().sort_index()
Movies_progress = Movies['release_year'].value_counts().sort_index()
```

```python
plt.figure(figsize=(14,7))

plt.plot(TVshows_progress.index, TVshows_progress.values, label='TV Shows')
plt.plot(Movies_progress.index, Movies_progress.values, label='Movies')

# below peice of code make marked line from end of the 2019 to starting of the 2021 and third line of code color the area
#---------------------------------------------------------------------
plt.axvline(2019, alpha=0.3, linestyle='--', color='r')
plt.axvline(2021, alpha=0.3, linestyle='--', color='r')
plt.axvspan(2019, 2021, alpha=0.2, color='r', label='Coronavirus')

#---------------------------------------------------------------------

plt.xticks(list(range(1925,2021,5)),fontsize=12)
plt.title('Content growth throughout history', fontsize=18)
plt.xlabel('Year', fontsize=16)
plt.ylabel('Amount of content', fontsize=16)

plt.yticks(fontsize=12)
plt.legend(prop={'size': 20})
plt.show()
```

```python
categories = list(s.split(' ') for s in df['listed_in'])
```

```python
from pandas.core.common import flatten
categories = list(flatten(categories))
len(categories)
# categories = np.array(categories)
# categories = np.unique(categories)
# categories
```

```python
import re
```

```python
categories = [re.sub(r'[^\w\s]','', element) for element in list(categories)]  
```

```python
np.unique(categories)
# len(categories)
```

```python
Counter(categories)
```

```python
# i will take only most 20 genre available
most_available_genre = Counter(categories).most_common(20)
```

```python
plt.figure(figsize=(16,8))
plt.title('Series/Movie available for a Genere',fontsize=20)
plt.xlabel('Count',fontsize=16)
plt.ylabel('Genre',fontsize=16)

plt.legend()
sns.countplot(y=categories)

```
