```python
import matplotlib.pyplot as plt
import seaborn as sns
data= pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')
data
```

```python
print(data.info())
print(data.describe())
```

```python
data.isnull().sum()
```

```python
#since we have more no missing data in director and cast we can drop
data_1=data.copy()
data_1=data_1.drop(['director','cast'],axis=1)
data_1
```

```python
# the show id and data_added is not going to affect the rating .so we can remove from data
data_1=data_1.drop(['show_id','date_added'],axis=1)
data_1
```

```python
data_2=data_1.copy()
data_2
```

```python
data_2['rating'].mode()[0]
```

```python
plt.pie(data['type'].value_counts(),labels=data_2['type'].value_counts().index)
plt.title('distribution of type')
plt.show()
```

```python
columns=data_2[['type','country','rating','release_year','listed_in']]
```

```python
for i in columns:
    plt.figure(figsize=(13,7))
    sns.countplot(y=columns[i],order=columns[i].value_counts().index[:20],palette='rocket_r')
    plt.title(i)
    plt.xticks(rotation=45,fontsize = 25)
    plt.yticks(fontsize = 20)
    plt.show()
```
