Firstly, I register the data into data variable

I changed Abnormal = 1 and Normal=0 for machine train because skleran works categorical type or integer type 

```python
data["class"] = [1 if each == "Abnormal" else 0 for each in data["class"]]
```

```python
data.head()
```

I created y and x variables for make train and test split 

```python
y=data["class"].values
x_data=data.drop(["class"],axis=1)
```

I did normalization here because of better correlation

```python
#normalization
x=(x_data-np.min(x_data))/(np.max(x_data)-np.min(x_data))
```

I split my data beacuse I trained my machine some data then I will correlate my test data 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

I trained my machine here and wanted that it can predict by x_test

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)
```

I learned here that how long make prediction errorless but in your opinion the machine make prediction better that acorrding to old prediction. We see

```python
print("{}, nn accuracy {}".format(3,knn.score(x_test,y_test)))
```

```python
score_list = []
for each in range(1,15):
    knn2=KNeighborsClassifier(n_neighbors=each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))

 #visualization   
plt.figure(figsize=(10,10))
plt.plot(range(1,15),score_list,color="black")
plt.legend()
plt.xlabel=('n_neigbors values')
plt.ylabel=('accuracy')
plt.show()
    
```
