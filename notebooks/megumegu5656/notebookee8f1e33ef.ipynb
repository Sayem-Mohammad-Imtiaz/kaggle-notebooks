```python
corr_mat = test.corr(method='pearson')
corr_mat
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(corr_mat)
plt.show()
```

```python
corr_mat["G3"]
```

```python
#G2:0.904868, G1:0.801468, failures:-0.360415,Medu:0.217147
from sklearn import linear_model
reg=linear_model.LinearRegression()
y=test[["G3"]]
x=test[["G1","G2","failures","Medu"]]
```

```python
reg.fit(x,y)
```

```python
print('回帰係数は',reg.coef_,'\n') 
print('切片は',reg.intercept_)
```

```python
plt.scatter(y,reg.predict(x))

# 散布図をプロットする
plt.xlabel('jissoku')
plt.ylabel('yosoku')
```

```python
reg.score(x,y)
```

```python
n = x.shape[0] #データ数
k = x.shape[1]+1 #説明変数の列数＋1(切片)
R1 = reg.score(x,y) #決定係数

R1_adj = 1-((n-1)/(n-k))*(1-R1)
R1_adj
```
