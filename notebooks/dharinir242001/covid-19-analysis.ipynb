```python
#Covid 19 Visualization

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
#This includes analysis of country wise and day wise covid 19 active cases, recovered, deaths and total number of cases.
```

```python
#Reading the necessary data and exploring the first five dataset values

country_wise = pd.read_csv("../input/corona-virus-report/country_wise_latest.csv")
day_wise = pd.read_csv("../input/corona-virus-report/day_wise.csv")
country_wise.head()
```

```python
day_wise.head()
```

```python
#Using seaborn let's analyze the scenario

import seaborn as sns
country_categorical = country_wise[['Confirmed','Recovered','Active','Deaths']]
sns.lineplot(data = country_categorical)
```

```python
#The above line plot indicates the country wise deaths, recovered, active and confirmed cases. It is seen that there is a sudden increase in the number of confirmed cases whereas the number of deaths is quite stable.
```

```python
day_categorical = day_wise[['Confirmed','Recovered','Active','Deaths']]
sns.lineplot(data = day_categorical)
```

```python
#This is the day wise visualization of the parameters.There is a gradual rise in the number of recovered as well as confirmed cases. The number of deaths is quite stable and is on the lower end.
```

```python
sns.kdeplot(data = country_wise['New deaths'],shade = True)
```

```python
#The number of new deaths in the country is seen to have sharp increase on certain days
```

```python
sns.kdeplot(data = day_wise['New deaths'],shade = True)
```

```python
#The day wise plot shows that there is a smooth fluctuation in the number of new deaths
```

```python
sns.heatmap(data = day_wise[['New cases','New deaths','New recovered']])
```

```python
#The heatmap gives a better visualization of the number of day wise new cases, deaths and recovered.
```

```python
sns.jointplot(x = country_wise['Active'],y = day_wise['Active'],kind = 'kde')
```

```python
#This joint plot shows a comparitive relation between day wise and country wise active cases. This would help in taking further necessary actions.
```
