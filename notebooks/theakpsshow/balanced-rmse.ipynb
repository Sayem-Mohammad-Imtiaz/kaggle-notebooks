# Handeling Null Values

```python
df['RainTomorrow'].fillna('Yes',inplace=True)
df['RainToday'].fillna('Yes',inplace=True)
df['Temp3pm'].fillna(method='ffill',inplace=True)
df['Temp9am'].fillna(method='ffill',inplace=True)
df['Cloud3pm'].fillna(method='bfill',inplace=True)
df['Cloud9am'].fillna(method='ffill',inplace=True)
df['Pressure3pm'].fillna(method='ffill',inplace=True)
df['Pressure9am'].fillna(method='ffill',inplace=True)
df['Humidity3pm'].fillna(method='ffill',inplace=True)
df['Humidity9am'].fillna(method='ffill',inplace=True)
df['WindSpeed3pm'].fillna(method='ffill',inplace=True)
df['WindSpeed9am'].fillna(method='ffill',inplace=True)
df['WindDir3pm'].fillna(method='ffill',inplace=True)
df['WindDir9am'].fillna(method='ffill',inplace=True)
df['WindGustSpeed'].fillna(method='ffill',inplace=True)
df['WindGustDir'].fillna(method='ffill',inplace=True)
df['MinTemp'].fillna(method='ffill',inplace=True)
df['MaxTemp'].fillna(method='ffill',inplace=True)
df['Rainfall'].fillna(method='ffill',inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.countplot(x='Temp3pm',data=df)
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
char_columns = ['Location','WindGustDir','WindDir9am','WindDir3pm','RainTomorrow','RainToday']
for i in char_columns:
    df[i] = le.fit_transform(df[i])
```

```python
df.drop('Date',axis=1,inplace=True)
```

```python
df.head()
```

```python
x = df.drop('RainTomorrow',axis=1)
y = df['RainTomorrow']
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(verbosity=0)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.33,random_state=0)
```

```python
xgb.fit(x_train,y_train)
```

```python
from xgboost import plot_importance
plot_importance(xgb)
```

```python
prediction = xgb.predict(x_test)
```

```python
final_output = []
for i in prediction:
    if i > 0.5:
        final_output.append('1')
    else:
        final_output.append('0')
```

```python
main_list = []
for i in final_output:
    if i == '0':
        main_list.append('No Rain')
    else:
        main_list.append('Rain')
```

```python
d = {'Rain_Prediction':main_list}
ddf = pd.DataFrame(d)
ddf.head()
```

```python
sns.displot(main_list)
```
