```python
mushrooms = pd.read_csv('/kaggle/input/mushroom-classification/mushrooms.csv')
```

```python
mushrooms.head()
```

```python
mushrooms.info()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
X = mushrooms.iloc[:, 1:]
```

```python
y = mushrooms['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=42)
```

```python
X_train.columns.values
```

```python
X_train_encoded = X_train.copy()
X_test_encoded = X_test.copy()
for col in X_train.columns:
    X[col] = le.fit_transform(X[col])
    X_train_encoded[col] = le.fit_transform(X_train[col])
    X_test_encoded[col] = le.transform(X_test[col])
```

```python
X_train_encoded.shape
```

```python
X_train_encoded
```

```python
X_test_encoded
```

```python
from sklearn.linear_model import LogisticRegression, SGDClassifier
```

```python
lr = LogisticRegression(max_iter=500)
```

```python
lr.fit(X_train_encoded, y_train)
```

```python
lr.score(X_test_encoded, y_test)
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
y_pred = lr.predict(X_test_encoded)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)
```

```python
conf_matrix
```

```python
import seaborn as sns
```

```python
sns.heatmap(conf_matrix, annot=True, cmap="YlGnBu")
```

```python
sgd = SGDClassifier()
```

```python
sgd.fit(X_train_encoded, y_train)
```

```python
sgd.score(X_test_encoded, y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params_sgd = {
    'penalty':['l2', 'l1', 'elasticnet'],
    'alpha':[0.0001, 0.001, 0.01]
}
```

```python
grid_sgd = GridSearchCV(sgd, params_sgd, cv=5)
```

```python
grid_sgd.fit(X_train_encoded, y_train)
```

```python
grid_sgd.best_params_
```

```python
grid_sgd.score(X_test_encoded, y_test)
```

```python
X.head()
```

```python
y_pred = grid_sgd.predict(X_test_encoded)
```

```python
output = pd.DataFrame({'Id':X_test_encoded.index,
                     'Class':y_pred})
```
