Birds Dataset

Print train and test acc for KNN, DTC and RF 

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../input/birds-bones-and-living-habits/bird.csv')
data
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
data
```

```python
data.groupby('type').size()
```

```python
X = data.drop('type', axis = 1)
Y = data['type']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size =0.4 ,random_state=2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, Y_train)
#Y_pred = knn.predict(X_test)
knn.score(X_test, Y_test)
```

```python
knn.score(X_train, Y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state = 0)
dtc.fit(X_train, Y_train)
dtc.score(X_test, Y_test)
```

```python
dtc.score(X_train, Y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=20, random_state=0)
rfc.fit(X_train, Y_train)
#Y_pred1 = rfc.predict(X_test)
rfc.score(X_test, Y_test)
```

```python
rfc.score(X_train, Y_train)
```

```python
# knn-train-0.9826989619377162
# knn-test-0.9758064516129032
# dtc-train-1.0
# dtc-test-0.9758064516129032
# rfc-train-1.0
# rfc-test-0.9516129032258065
```

Run cross validation on the Birds Dataset to eliminate overfitting . 

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(rfc, X_train, Y_train, cv=5)
scores.mean()
```

```python
scorestest = cross_val_score(rfc, X_test, Y_test, cv=5)
scorestest.mean()
```

```python
scores1 = cross_val_score(dtc, X_train, Y_train, cv=5)
scores1.mean()
```

```python
scores2 = cross_val_score(knn, X_train, Y_train, cv=5)
scores2.mean()
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=5)
kscore = cross_val_score(rfc, X_train, Y_train, scoring='accuracy', cv=kf, n_jobs=-1)
kscore
kscore.mean()
```

```python
kscoretest = cross_val_score(rfc, X_test, Y_test, scoring='accuracy', cv=kf, n_jobs=-1)
kscoretest.mean()
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5)
skfscore = cross_val_score(rfc, X_train, Y_train, scoring='accuracy', cv=skf, n_jobs=-1)
skfscore.mean()
```

```python
skfscoretest = cross_val_score(rfc, X_test, Y_test, scoring='accuracy', cv=skf, n_jobs=-1)
skfscoretest.mean()
```
