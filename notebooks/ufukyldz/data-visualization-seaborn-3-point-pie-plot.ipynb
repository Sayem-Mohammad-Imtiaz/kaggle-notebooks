DATA 1 ( PERCENTAGE PEOPLE BELOW POVERTY LEVEL ) 

```python
dataPoverty = pd.read_csv("/kaggle/input/percentage-people-below-poverty-level/PercentagePeopleBelowPovertyLevel.csv" , encoding = "windows-1252")
```

```python
dataPoverty.head()
```

```python
dataPoverty.info()
```

```python
dataPoverty.value_counts()
```

UNDEFINDED VARIABLES

```python
dataPoverty["poverty_rate"].value_counts()
```

NAN VALUES

```python
dataPoverty["poverty_rate"].isna().sum()
```

```python
dataPoverty.replace("-" , 0.0 , inplace = True)

dataPoverty["poverty_rate"].value_counts()
```

```python
dataPoverty.poverty_rate = dataPoverty.poverty_rate.astype(float)

dataPoverty.info()
```

```python
places = dataPoverty["Geographic Area"].unique()

places
```

```python
povertyRatio = []

for place in places:
    x = dataPoverty[dataPoverty["Geographic Area"] == place]
    average = sum(x.poverty_rate) / len(x)
    povertyRatio.append(average)
    
povertyRatio
```

```python
dfPoverty = pd.DataFrame({
    "places" : places,
    "poverty" : povertyRatio
    
})

indexPoverty = (dfPoverty["poverty"].sort_values(ascending = True)).index.values
newIndexPoverty = dfPoverty.reindex(indexPoverty)
```

```python
newIndexPoverty
```

COMPRESSING THE RATIOS OF TWO DATA BETWEEN 0 AND 1

```python
newIndexPoverty["poverty"] = newIndexPoverty["poverty"] / max(newIndexPoverty["poverty"])
newIndexHS["grad"] = newIndexHS["grad"] / max(newIndexHS["grad"])
```

```python
dataLast = pd.concat([ newIndexPoverty , newIndexHS["grad"]] , axis = 1)
dataLast.sort_values("poverty" , inplace = True) 
dataLast
```
