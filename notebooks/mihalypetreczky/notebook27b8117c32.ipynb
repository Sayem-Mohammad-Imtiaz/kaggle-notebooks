Basic statistics: mean, standard deviation, correlation

mean: $m_x=\frac{1}{n} \sum_{i=1}^{n} x_i$

standard deviation: $\sigma_{x} =\sqrt{\frac{1}{n-1} \sum_{i=1}^{n} (x_i-m_x)^2}$


correlation: $cor_{x,y}=\frac{\frac{1}{n-1} \sum_{i=1}^{n} (x_i-m_x)(y_i-m_y)}{\sigma_x \sigma_y}$.

Machine learning, applying linear regression.

Splitting the data into training and testing sets

```python
#import from the library the function to split data
from sklearn.model_selection import train_test_split
```

```python
X=df['YearsExperience'] #feature, predictor
Y=df['Salary']  #variable to be predicted, labels
```

```python
result=train_test_split(X,Y,test_size=0.2) #20% of data points for testing
```

```python
len(result)
```

```python
X_train = result[0] #Predictors from training set
X_test  = result[1] #Predictors from testing set
Y_train = result[2] #Labels of the training set 
Y_test  = result[3] #Labels for the testing set
```

```python
type(X_train)
```

```python
type(Y_train)
```

```python
X_train.shape
```

```python
Y_train.shape
```

Machine learning: learning the model

Step 1: choose a model class, we choose linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()  #create variable containing an
                         #unitialized linear regression model
```

Fit the model lr to data

```python
#Transform X_train to n x 1 np.array
```

```python
xt=X_train.to_numpy() #convert X_train series to np.array
                      #xt is a 1D array (just a sequence)
                      #We want xt to be n x 1 array, n is the number of
                      #data points in the training set
xt=xt.reshape(-1,1)
#the same as xt=X_train.to_numpy().reshape(X_train.shape[0],1)
```

```python
lr.fit(xt,Y_train) #fitting your model to training data.
```

Now, lr contains the trained model. We have to test using the test set

```python
Ypred=lr.predict(X_test.to_numpy().reshape(-1,1))
#Use test data to predict the labels, reshape test data in X_test to make
#a n x 1 array (n rows, 1 column, n is the number of elements of X_test)
```

```python
Ypred-Y_test
```

```python
#Difficult to decide if Ypred-Y_test is large or not
#Option; visual check
plt.plot(range(X_test.shape[0]),Y_test,'r', range(X_test.shape[0]),Ypred, 'b')
```

```python
#Relative norm of the error:
np.linalg.norm(Y_test-Ypred)/np.linalg.norm(Y_test)
```

Relative error: 8%, quite good

```python
#Mean square error 
np.sqrt(np.linalg.norm(Y_test-Ypred)**2/(Y_test.shape[0]-1))
```
