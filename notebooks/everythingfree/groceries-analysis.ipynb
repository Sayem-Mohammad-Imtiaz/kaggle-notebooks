```python
import pandas as pd
from matplotlib import pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
df=pd.read_csv(r"../input/groceries/groceries - groceries.csv")
df.head()
```

```python
df.shape
```

```python
df.drop(columns=['Item(s)'],inplace=True)
df.isnull().sum().all()
```

```python
te=TransactionEncoder()
df=pd.DataFrame(te.fit_transform([df.iloc[i].dropna().tolist() for i in range(len(df))]),columns=te.columns_)
df
```

```python
df.sum().sort_values(ascending=False).plot(kind='bar',figsize=(50,5),ylabel='No Of Occurance',xlabel='Items')
```

```python
filters=apriori(df,.05,True)
```

```python
filters.set_index('itemsets').sort_values('support',ascending=False).plot(kind='bar',figsize=(20,5),ylabel='Support')
```

```python
association_rules(filters,metric='lift',min_threshold=1).sort_values('lift',ascending=False)
```

> **As We Can See Whole Milk Is Tightly Associated With All Items.** 
<br>
This may be because *Whole Milk* is the most selling item in the grocery store.
