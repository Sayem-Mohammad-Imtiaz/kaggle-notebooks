considering only featues with higtest postive & negativ corealtion with the house price of unit area

```python
X=df[["X3 distance to the nearest MRT station","X4 number of convenience stores","X5 latitude","X6 longitude"]]
y=df["Y house price of unit area"]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y, test_size=0.4,random_state=5)
```

```python
xtest.shape
```

```python
ytest.shape
```

```python
lr=LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
lr.coef_
```

```python
coff_df=pd.DataFrame(lr.coef_,X.columns,columns=["corelationcoff"])
coff_df
```

keeping other fetures as is one unit incrase in distance to the nearest MRT station will reduce house price by 0.005 
keeping other fetures as is one unit incrase in number of convenience stores is associated with increase of house price by 0.8097


```python
y_pred=lr.predict(xtest)
```

```python

sns.distplot(y_pred)
from sklearn import metrics
```

```python
print("Mean Squared Error",metrics.mean_squared_error(ytest,y_pred))
print("Root Mean Squared Error",np.sqrt(metrics.mean_squared_error(ytest,y_pred)))
print("Mean Absolute Error", metrics.mean_absolute_error(ytest, y_pred))
print("{0}% of data point are close to regression line (coff odf det)".format(100*(metrics.r2_score(ytest,y_pred))))
```
