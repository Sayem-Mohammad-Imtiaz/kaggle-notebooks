First, read the given csv file via pandas. The dataframe (=df) of the day will only include the countries and their respective happiness ranks / scores . Let's ditch the rest!

```python
df = pd.read_csv('../input/2017.csv')
df = df[['Country', 'Happiness.Rank', 'Happiness.Score']]
```

Now begins the only little hassle of this project: The country data is provided in full names ("France") and not in the three digit-code ("FRA") which we'll later need as an input for our world map. 
For that concern, we'll need pycountry to define a function *lookup* that goes through all the names and checks if there's a corresponding ISO code.  Note that for each country, we'll check for matches in the name field ("Germany") as well as in the official name field ("Federal Republic of Germany")

```python
countries= df['Country'].values

def lookup(countries):
    result = []
    for i in range(len(countries)):
        try:
            result.append(pycountry.countries.get(name=countries[i]).alpha_3)
        except KeyError:
            try:
                result.append(pycountry.countries.get(official_name=countries[i]).alpha_3)
            except KeyError:
                result.append('undefined')
    return result
```

Use this function to fetch the codes and append them to our dataframe. Then get rid of all the unidentifyable countries (i.e. those countries for which no ISO code was found.

```python
codes=lookup(countries)
df['Codes']=codes
df=df[~df.Codes.isin(['undefined'])]
```

Next up, we insert the data into the data file for plotly. Remember that a plotly plot basically consists of a json that lays behind it. That json collects all data and layout information.

```python
data = [ dict(
    type = 'choropleth',
    locations = df['Codes'],
    z = df['Happiness.Score'],
    text = df['Country'],
   colorscale = [[0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"], \
                 [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"]],
   autocolorscale = False,
   reversescale = True,
    marker = dict(
       line = dict (
           color = 'rgb(180,180,180)',
           width = 0.5
       ) ),
    colorbar = dict(
       autotick = False,
       title = 'Happiness Score'),
) ]


layout = dict(
    title = 'Happiness Scores',
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )
    )
)
```

Eventually, we can plot it.  :) 

```python
fig = dict( data=data, layout=layout )
iplot(fig, validate=False)
```
