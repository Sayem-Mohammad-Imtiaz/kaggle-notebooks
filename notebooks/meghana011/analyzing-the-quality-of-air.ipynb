```python
df = pd.read_csv('../input/india-air-quality-data/data.csv',encoding ='cp1252')
```

```python
df.isnull().mean().round(5)*100
```

```python
df.head()
```

```python
df['agency'].value_counts()
```

```python
df['location'].fillna(df['location'].mode()[0], inplace=True)
df['type'].fillna(df['type'].mode()[0], inplace=True)
df['so2'].fillna(df['so2'].mean(),inplace=True)
df['no2'].fillna(df['no2'].mean(),inplace=True)
df['rspm'].fillna(df['rspm'].mean(),inplace=True)
df['date'].fillna(df['date'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.drop(['stn_code','agency','spm','location_monitoring_station','pm2_5','sampling_date'], axis=1,inplace=True)
```

```python
df['pollution'] = df['so2'] + df['no2'] + df['rspm']/3
```

```python
df[['state','pollution']].groupby(['state']).agg('mean').sort_values('pollution',ascending=False).head()
```

```python
df[['location','pollution']].groupby(['location']).agg('mean').sort_values('pollution',ascending=False).head()
```

```python
df['type'] = df['type'].replace(['Industrial Area', 'Industrial Areas'], 'Industrial')
df['type'] = df['type'].replace(['Sensitive Areas'], 'Sensitive')
df['type'] = df['type'].replace(['Residential and others'], 'Residential')
df['type'].str.split(',')[0]
```

```python
df['type'] = df['type'].str.split(', ')
df['type1'] = df['type'].apply(lambda x:x[0])
df['type2'] = df['type'].apply(lambda x:x[1] if len(x)>1 else x[0])
df[['type','type1','type2']].head()
```

```python
df[['type1','pollution']].groupby(['type1']).agg('mean').sort_values('pollution',ascending=False).head()
```

```python
df[['location','type1','so2']].groupby(['location','type1']).agg('mean').sort_values('so2',ascending=False).head(10)
```

```python
df[['location','type1','no2']].groupby(['location','type1']).agg('mean').sort_values('no2',ascending=False).head(10)
```

```python
df[(df['type1'] != 'RIRUO') | (df['type1'] != 'Rural and other Areas')][['location','type1',
                                                                         'pollution']].groupby(['location','type1']).agg('mean').sort_values('pollution',ascending=False).head(3)
```

```python
df[(df['type1'] == 'RIRUO') | (df['type1'] == 'Rural and other Areas')][['location','type1',
                                                                         'pollution']].groupby(['location','type1']).agg('mean').sort_values('pollution',ascending=False).head(3)
```
