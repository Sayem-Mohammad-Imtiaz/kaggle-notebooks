```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(rc={'figure.figsize':(10, 8)}); # you can change this if needed
```

Загрузите данные. Они находятся в директории input.

```python
df = pd.read_csv("../input/bigml_59c28831336c6604c800002a.csv")
df.head()
```

Каково распределение абонентов по целевому признаку **churn**? Назовите процент лояльных абонентов.

```python
sns.countplot(x = "churn", data = df)
loyal_users = sum(df["churn"] == True)
not_loyal_users = sum(df["churn"] == False)
print(not_loyal_users/(loyal_users + not_loyal_users) * 100)
```

Жители какого штата говорят по телефону больше всего? меньше всего? Обратите внимание, что нас интересуют все виды звонков.

```python
df1 = df.groupby("state")["total day calls", "total eve calls", "total night calls", "total intl calls", "customer service calls"].sum().sum(axis = 1).sort_values(ascending = False)
print(df1.iloc[:1], df1.iloc[-1:])
```

Верно ли, что люди, совершающие много международных звонков, приобретают **international plan**? Не забывайте о визуализации информации.

```python
print(df.groupby("international plan")["total intl calls"].sum())
sns.boxplot(x = "international plan", y = "total intl calls", data=df)
plt.show()
```

Как связаны наличие голосовой почты и международного плана с лояльностью абонентов? Проверьте статистическую значимость этих связей.

```python
print(df.groupby(["international plan", "voice mail plan"])["churn"].count())
```

Что можно сказать о взаимосвязи количества обращений в колл-центр и лояльности? Не забывайте о визуализации информации.

```python
print(df.groupby("churn")["customer service calls"].sum())
sns.boxplot(x = "churn", y = "customer service calls", data = df)
plt.show()
# Mostly not loyal users have more calls to the customer service
```

Какие из числовых признаков имеют наибольшую степень корреляции? Почему?

```python
df2 = df[["account length", "number vmail messages", "total day minutes", "total day calls", "total day charge", "total eve minutes", "total eve charge", "total night minutes", "total night calls", "total night charge", "total intl minutes", "total intl calls", "total intl charge", "customer service calls"]]
df2.corr()
sns.heatmap(df2.corr())
```

```python
from scipy.stats import pearsonr, spearmanr, kendalltau
r = pearsonr(df['total night calls'], df['total night minutes'])
print('Pearson correlation:', r[0], 'p-value:', r[1])
```

```python
df["international plan"] = df["international plan"].map({"yes":1, "no": 0})
df["voice mail plan"] = df["voice mail plan"].map({"yes":1, "no":0})
r = pearsonr(df['international plan'], df['voice mail plan'])
print('Pearson correlation:', r[0], 'p-value:', r[1])
```

```python
r = spearmanr(df['international plan'], df['voice mail plan'])
print('Spearman correlation:', r[0], 'p-value:', r[1])
```
