```python
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer('bert-base-nli-mean-tokens')
```

```python
mod = SentenceTransformer('paraphrase-distilroberta-base-v1')
```

```python
title= df.title.values.tolist()
```

```python
title_embeddings= model.encode(title)
```

```python
query = '3 Generations'

queries = [query]
query_embeddings = model.encode(queries)

# Find the closest 10 sentences of the corpus for each query sentence based on cosine similarity
number_top_matches = 10

print("Semantic Search Results")

for query, query_embedding in zip(queries, query_embeddings):
    distances = scipy.spatial.distance.cdist([query_embedding], title_embeddings, "cosine")[0]

    results = zip(range(len(distances)), distances)
    results = sorted(results, key=lambda x: x[1])

    print("\n\n=========================================================================================================================\n\n")
    print("Query:", query)
    print("\n\n=========================================================================================================================\n\n")
    print("\nTop 10 most similar sentences in corpus:")
    print("\n\n=========================================================================================================================\n\n")

    for idx, distance in results[0:number_top_matches]:
        print(title[idx].strip(), "(Cosine Score: %.4f)" % (1-distance))
```
