```python
df = pd.read_csv('/kaggle/input/ip-network-traffic-flows-labeled-with-87-apps/Dataset-Unicauca-Version2-87Atts.csv', nrows=10000)
```

```python
df = df.dropna()
feature_set = ['Average.Packet.Size','Destination.Port','Destination.IP'] #,'Source.IP']

from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)
cat_encoded = encoder.fit_transform(df[feature_set[1:]])
print(type(cat_encoded), cat_encoded.shape)
cont_data = df[feature_set[0]].to_numpy().reshape(-1,1) 
print(type(cont_data), cont_data.shape)
x_data = np.append(cat_encoded, cont_data, axis=1)
print(x_data.shape)
y_labels = df['ProtocolName']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_data, y_labels, test_size=0.3, random_state=42)
```

```python
from sklearn import tree

# Create a tree object
classifier = tree.DecisionTreeClassifier()
# Ask the tree to fit the data to the labels!  That's it!
classifier = classifier.fit(X_train, y_train)

from sklearn.metrics import confusion_matrix
results = confusion_matrix(y_test, classifier.predict(X_test))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, classifier.predict(X_test))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(20,10))
sns.heatmap(results, fmt="d")
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```
