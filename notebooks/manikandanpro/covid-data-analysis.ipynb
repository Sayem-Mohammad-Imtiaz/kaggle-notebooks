**About the Dataset**

* This dataset contains latest Covid-19 India state-wise data as on August 18, 2021. This dataset can be used to analyze covid condition in India.

* This dataset is great for Exploratory Data Analysis

* Attribute Information

* State/UTs - Names of Indian States and Union Territories.

* Total Cases - Total number of confirmed cases

* Active - Total number of active cases

* Discharged - Total number of discharged cases

* Deaths - Total number of deaths

* Active Ratio (%) - Ratio of number of active cases to total cases

* Discharge Ratio (%) - Ratio of number of discharged cases to total cases

* Death Ratio (%) - Ratio of number of deaths to total cases**

**Gathering the Information from the Datasets**

```python
df=pd.DataFrame(pd.read_csv("../input/d/anandhuh/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv"))
df
```

**Here below we are going to compare all the features with the particular state which is headed above which helps us in identifying the current status of the covid of that particlar State/UTs.**

# **Analysis of State/UTs with the Total Cases**

```python
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i, y[i], y[i], ha = 'center',color='red',fontsize=15)
figure(figsize=(25,15),dpi=80)
ax = plt.axes()
ax.set_facecolor("black")
a=plt.plot(df['State/UTs'],df['Total Cases'],color='white',linewidth=1)
b=plt.scatter(df['State/UTs'],df['Total Cases'],color='red',s=30)
addlabels(df['State/UTs'],df['Total Cases'])
plt.xlabel("State/UTs",color='blue',fontsize=20)
plt.ylabel("Total Cases",color='blue',fontsize=20)
plt.xticks(rotation=90,fontsize=15)
plt.title("Analysis of State/UTs with the Total cases statistics",color='blue',fontsize=25)
plt.show()
```
