# **Comparative Analysis of Age and Number of persons on that Particular age group**

```python
figure(figsize=(16,8),dpi=80)
ax = plt.axes()
ax.set_facecolor("black")
Age_factor=pd.DataFrame(a['Age'].value_counts())
plt.scatter(Age_factor.index,Age_factor['Age'],color='white',facecolor='red',s=350)
plt.bar(Age_factor.index,Age_factor['Age'],color='blue',width=0.1)
plt.title('Comparative Analysis of Age and Number of persons on that Particular age group',color='blue',fontsize=20)
plt.xticks(size=15,color='red')
plt.xlabel('Age',size=15,color='blue')
plt.yticks(size=15,color='red')
plt.ylabel('Count of grouping of Age',size=15,color='blue')
plt.show()
```

**1.   31 and 29 has the highest count compared to others**

**2.   26-35 years is the average highest count of age factor decision**
