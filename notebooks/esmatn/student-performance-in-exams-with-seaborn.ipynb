## Loading Data and Explanation of Features

```python
#to read data
df = pd.read_csv("../input/students-performance-in-exams/StudentsPerformance.csv")
```

```python
#Convert object race/ethnicity values into integer values
race = {'group A':1, 'group B':2,'group C':3, 'group D':4, 'group E':5} 
df['race/ethnicity'] = [race[item] for item in df['race/ethnicity']] 
#to convert int into object
#race = {1:'group A', 2:'group B',3: 'group C', 4: 'group D', 5: 'group E'} 
#df['race/ethnicity'] = [race[item] for item in df['race/ethnicity']] 
```

```python
# Convert object gender values into integer values
gender = {'male':1,'female':2} 
df.gender = [gender[item] for item in df.gender]
#to convert int into object
#gender = {1:'male',2:'female'} 
#df.gender = [gender[item] for item in df.gender]
```

```python
gender_list = list(df['gender'].unique())
race_list = list(df['race/ethnicity'].unique())
```

## Visualization with Seaborn 

<a id=31></a>
* ## Bar Plot 

```python
#Gender--Math Score ratio
gender_math_ratio = []
for i in gender_list:
    x = df[df['gender']==i]
    gender_math_rate = sum(x['math score'])/len(x)
    gender_math_ratio.append(gender_math_rate)
data = pd.DataFrame({'gender_list': gender_list, 'gender_math_ratio':gender_math_ratio})
new_index = (data['gender_math_ratio'].sort_values(ascending=False)).index.values
sorted_data = data.reindex(new_index)

#visualization
plt.figure(figsize=(5,5))
ax = sns.barplot(x=sorted_data['gender_list'], y=sorted_data['gender_math_ratio'])
plt.xticks(rotation=0)
plt.xlabel('Genders')
plt.ylabel('Math scores')
plt.title('Math Scores Rate Given Gender')
plt.show()
```

```python
#Gender--Writing Score ratio
gender_writing_ratio = []
for i in gender_list:
    x = df[df['gender']==i]
    gender_writing_rate = sum(x['writing score'])/len(x)
    gender_writing_ratio.append(gender_writing_rate)
data = pd.DataFrame({'gender_list': gender_list, 'gender_writing_ratio':gender_writing_ratio})
new_index = (data['gender_writing_ratio'].sort_values(ascending=True)).index.values
sorted_data2 = data.reindex(new_index)

#visualization
plt.figure(figsize=(5,5))
ax = sns.barplot(x=sorted_data2['gender_list'], y=sorted_data2['gender_writing_ratio'])
plt.xticks(rotation=0)
plt.xlabel('Genders')
plt.ylabel('Writing scores')
plt.title('Writing Scores Rate Given Gender')
plt.show()
```

```python
#Gender--Reading Score ratio
gender_reading_ratio = []
for i in gender_list:
    x = df[df['gender']==i]
    gender_reading_rate = sum(x['reading score'])/len(x)
    gender_reading_ratio.append(gender_reading_rate)
data = pd.DataFrame({'gender_list': gender_list, 'gender_reading_ratio':gender_reading_ratio})
new_index = (data['gender_reading_ratio'].sort_values(ascending=True)).index.values
sorted_data3 = data.reindex(new_index)

#visualization
plt.figure(figsize=(5,5))
ax = sns.barplot(x=sorted_data3['gender_list'], y=sorted_data3['gender_reading_ratio'])
plt.xticks(rotation=0)
plt.xlabel('Genders')
plt.ylabel('Reading scores')
plt.title('Reading Scores Rate Given Gender')
plt.show()
```

```python
#Race--Math Score ratio
race_math_ratio = []
for i in race_list:
    x = df[df['race/ethnicity']==i]
    race_math_rate = sum(x['math score'])/len(x)
    race_math_ratio.append(race_math_rate)
data = pd.DataFrame({'race_list': race_list, 'race_math_ratio':race_math_ratio})
new_index = (data['race_math_ratio'].sort_values(ascending=False)).index.values
sorted_data4 = data.reindex(new_index)

#visualization
plt.figure(figsize=(5,5))
ax = sns.barplot(x=sorted_data4['race_list'], y=sorted_data4['race_math_ratio'])
plt.xticks(rotation=0)
plt.xlabel('Race/Ethnicities')
plt.ylabel('Math scores')
plt.title('Math Scores Rate Given Race/Ethnicities')
plt.show()
```

```python
sns.countplot(data=df,x='test preparation course',hue='gender')
plt.show()
```

```python
sorted_data['gender_math_ratio'] = sorted_data['gender_math_ratio']/max( sorted_data['gender_math_ratio'])
sorted_data2['gender_writing_ratio'] = sorted_data2['gender_writing_ratio']/max( sorted_data2['gender_writing_ratio'])

#concatenate these two values
data = pd.concat([sorted_data,sorted_data2['gender_writing_ratio']],axis=1)

data.sort_values('gender_list',inplace=True)
```
