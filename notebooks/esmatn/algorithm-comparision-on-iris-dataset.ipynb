```python
#Load Data
data = pd.read_csv("../input/iris-flower-dataset/IRIS.csv")
data
```

```python
data.describe()
```

```python
# split data
X = data.iloc[:,:4].values 
y = data.iloc[:,4:].values 
```

```python
# label encoding
species = data.iloc[:,4:].values
from sklearn import preprocessing
#önce tr, fr ve us yi 0 1 ve 2 haline çeviriyoruz
le=preprocessing.LabelEncoder()
species[:,0] = le.fit_transform(data.iloc[:,4:])
y = species
y=y.astype('int')
```

```python
# plots
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5

plt.figure(1, figsize=(8,6))
plt.clf()

# 2D plot
plt.scatter(X[:,0], X[:,1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

# 3D plot the training points -- it's better to see this data in 3D
fig = plt.figure(2, figsize=(8,6))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(X[:,0], X[:,1], X[:,2], c=y, cmap=plt.cm.Set1, edgecolor='k', s=40)
ax.set_title("Iris Dataset")
ax.set_xlabel("First Spec")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("Second Spec")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("Third Spec")
ax.w_zaxis.set_ticklabels([])

plt.show()
```

```python
# train-test-split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.33, random_state=0)

# scaling dataset
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)
#y_test = y_test.reshape(-1,1)
```

```python
# Algorithms
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(random_state = 0)
log_reg.fit(X_train, y_train)

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=6, metric='minkowski')
knn.fit(X_train,y_train)

from sklearn.svm import SVC
svc = SVC(kernel='rbf')
svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)

from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion = 'entropy')
dtc.fit(X_train,y_train)

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=10, criterion = 'entropy')
rfc.fit(X_train,y_train)
```

```python
# confusion matrices
y_pred = log_reg.predict(X_test)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,y_pred)
print("Logistic Regression")
print(cm)

y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print("KNN")
print(cm)

y_pred = gnb.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print('SVC')
print(cm)

y_pred = dtc.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print('GNB')
print(cm)

y_pred = rfc.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print('DTC')
print(cm)

cm = confusion_matrix(y_test,y_pred)
print('RFC')
print(cm)
```
