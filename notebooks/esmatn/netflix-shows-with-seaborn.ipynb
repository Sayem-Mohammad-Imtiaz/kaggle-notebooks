```python
data["show_duration"] = ["high" if any("Season" in s for s in data.duration) else "low" for s in data.duration]
data.loc[:10,["show_duration","duration"]]
```

```python
#threshold(bir eşik değeri, burada ortalama yıl değerini hesapladım) önce yayınlanan programlar için eski, sonra yayınlanan programlar için yeni yazmak:
threshold = sum(data.release_year)/len(data.release_year)

year_mean = int(threshold) #float değeri integera çevirmek
#print(year_mean)#2013

data["release"] = ["old" if i<year_mean else "new" for i in data.release_year]
data.loc[:10,["release","release_year"]]
```
