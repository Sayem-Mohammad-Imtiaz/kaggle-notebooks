## Prepare the data for analysis

```python
file_path = "../input/pokemon/Pokemon.csv"

#read_csv() : make dataframe from pokemon file 
pokemon_df = pd.read_csv(file_path)

print(pokemon_df.info())

pokemon_df.head(5)
```

```python
pokemon_df[pokemon_df['Legendary']==True].head(5)  #Showing the legendary pokemons
```

```python
# The name is duplicated in front of Mega, so remove it.
pokemon_df['Name'] = pokemon_df['Name'].str.replace(".*(?=Mega)", "")
pokemon_df[pokemon_df['Legendary']==True].head(5) 
```

```python
#some values in TYPE2 are empty and thus they have to be filled or deleted
pokemon_df['Type 2'].fillna(pokemon_df['Type 1'], inplace=True) #fill NaN values in Type2 with corresponding values of Type

pokemon_df.head(5)
```

## Feature of Pokémon Data

Before doing the analysis, let's look at the overall feature of the data.

```python
pokemon_df.describe()
```

```python
stat_col = list(pokemon_df.columns[5:11])
print(stat_col)
```

```python
pokemon_df.groupby("Type 1")["#"].count()
```

```python
pokemon_df[pokemon_df["Legendary"] == True].groupby("Type 1")["#"].count()
```

### Correlation of stats

Attack is most often associated with defense. There are relations in order of HP and special attack.

Defence is most often associated with special defenses, followed by attacks and less associated with the other.

Special attack is often associated with special defense.

Speed is associated with special attack.


```python
stat_corr = pokemon_df[stat_col].corr(method = 'pearson')
print(stat_corr)
```

```python
sns.set(font_scale=1.5)
sns.heatmap(stat_corr.values,
            cbar = True,
            annot = True,
            fmt = '.2f',
            annot_kws = {'size': 15},
            yticklabels = stat_col,
            xticklabels = stat_col)

plt.show()

sns.set(style = 'whitegrid', context='notebook')
sns.pairplot(pokemon_df[stat_col], height = 2.5)

plt.show()
```

## Checking Boxplots

```python
pokemon_df.boxplot(column=stat_col)
```

## Top 10 Pokémon with the Highest Stats

```python
plt.figure(figsize=(20,14))
sns.barplot(x='Total', y='Name', data=pokemon_df, order=pokemon_df.sort_values('Total',ascending=False).iloc[:20].Name)
sns.despine()
plt.show()
```

## Distribution by Stat

```python
bins=range(0,200,20) #they act as containers
plt.hist(pokemon_df["Attack"],bins,histtype="bar",rwidth=1.2,color='#0ff0ff') #hist() is used to plot a histogram
plt.xlabel('Attack') #set the xlabel name
plt.ylabel('Count') #set the ylabel name
plt.plot()
plt.axvline(pokemon_df['Attack'].mean(),linestyle='dashed',color='red') #draw a vertical line showing the average Attack value
plt.show()
```

```python
bins=range(0,200,20) #they act as containers
plt.hist(pokemon_df["Defense"],bins,histtype="bar",rwidth=1.2,color='#0ff0ff') #hist() is used to plot a histogram
plt.xlabel('Defense') #set the xlabel name
plt.ylabel('Count') #set the ylabel name
plt.plot()
plt.axvline(pokemon_df['Defense'].mean(),linestyle='dashed',color='red') #draw a vertical line showing the average Attack value
plt.show()
```

```python
bins=range(0,200,20) #they act as containers
plt.hist(pokemon_df["Speed"],bins,histtype="bar",rwidth=1.2,color='#0ff0ff') #hist() is used to plot a histogram
plt.xlabel('Speed') #set the xlabel name
plt.ylabel('Count') #set the ylabel name
plt.plot()
plt.axvline(pokemon_df['Speed'].mean(),linestyle='dashed',color='red') #draw a vertical line showing the average Attack value
plt.show()
```

## Stat by type

```python
plt.subplots(figsize = (20,10))
plt.title('Attack by Type1')
sns.violinplot(x = "Type 1", y = "Attack",data = pokemon_df)
plt.ylim(0,200)
plt.show()
```

```python
plt.subplots(figsize = (20,10))
plt.title('Sp. Attack by Type1')
sns.violinplot(x = "Type 1", y = "Sp. Atk",data = pokemon_df)
plt.ylim(0,200)
plt.show()
```
