```python
pokemon = pd.read_csv('../input/Pokemon.csv')
pokemon.describe().transpose()
```

```python
#Count Pokemon Types
pokemonTypes = pokemon[['#', 'Type 1']].groupby('Type 1').count()
orderedTypes1 = pokemonTypes.sort_values('#').reset_index()['Type 1']

sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))
sns.countplot(y ='Type 1', data = pokemon, order=orderedTypes1, orient='h').set(ylabel='Pokemon Types', title = 'Count of Pokemon Types')
```

```python
#HP compared to Pokemon Type 1
pokemonStats = pokemon[['Type 1', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']]

plt.figure(figsize=(10, 16))
sns.violinplot(x='HP', y='Type 1', data = pokemonStats, orient = 'h')
```

```python
#Have pokemon designers been consitent in creating appropriate distributions for HP
pokemonHpTrend = pokemon[['Type 1', 'HP', 'Generation']]
pokemonHpTrend['GenerationOne'] = np.where(pokemonHpTrend['Generation']==1, 1, 0)

plt.figure(figsize=(10, 16))
sns.violinplot(x='HP', y='Type 1', hue='GenerationOne', data = pokemonHpTrend, split=True, orient='h', inner='quart' )
```

```python
#Trying out multiple plots in the same figure
#thanks for the tutorial iammax2

fig, axs = plt.subplots(4, 4, figsize = (10,10))

ax1 = plt.subplot2grid((4, 4), (0, 0), rowspan=2, colspan=2)
ax2 = plt.subplot2grid((4, 4), (0, 2), rowspan=2, colspan=2)
ax3 = plt.subplot2grid((4, 4), (2, 0), rowspan=2, colspan=2)
ax4 = plt.subplot2grid((4, 4), (2, 2), rowspan=2, colspan=2)

ax1.set_title("Sp. Atk Vs. Hp")
ax2.set_title("Sp. Atk Vs. Attack")
ax3.set_title("Sp. Atk Vs. Defense")
ax4.set_title("Sp. Atk Vs. Speed")

fig.tight_layout(pad = 0.5, w_pad = 4.0, h_pad = 4.0)

sns.kdeplot(pokemon['Sp. Atk'], pokemon['HP'], shade = True, cmap = 'ocean', ax=ax1)
sns.kdeplot(pokemon['Sp. Atk'], pokemon['Attack'], shade = True, cmap = 'plasma', ax=ax2)
sns.kdeplot(pokemon['Sp. Atk'], pokemon['Defense'], shade = True, cmap = 'rainbow', ax=ax3)
sns.kdeplot(pokemon['Sp. Atk'], pokemon['Speed'], shade = True, cmap = 'summer', ax=ax4)

sns.set(style = 'whitegrid')
sns.despine(left=True)
sns.set_context('paper')

plt.show()

#Choose the most obnoxious colors. 
```
