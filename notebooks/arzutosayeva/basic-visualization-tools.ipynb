```python
#Lets import dataset
data=pd.read_csv('/kaggle/input/videogamesales/vgsales.csv')
```

```python
#in order to use our data properly without missing original version we copy it
df=data.copy()
```

```python
df.info()
```

```python
df.isnull().values.any()
```

```python
df.isnull().sum()
```

```python
#Lets replace null values with 0
df['Year'].fillna(0,inplace=True)
```

```python
df['Publisher'].fillna(0,inplace=True)
```

```python
#lets see only object values
cat_var=df.select_dtypes(include=['object'])
```

```python
cat_var
```

```python
cat_var['Platform'].value_counts()
```

```python
#from the graph below we can see how video game NA_Sales changed in terms of genres
ax=sns.catplot(x='Genre',y='NA_Sales',data=df,height=4,aspect=2);

ax.set_xticklabels(rotation=30)
plt.show()
```

```python
#If we add one more variable we see our graph shows also specially the game genres NA_sales in terms of platforms
ax=sns.catplot(x='Genre',y='NA_Sales',hue='Platform',data=df,height=4,aspect=2)
ax.set_xticklabels(rotation=30)
plt.show()
```

```python
#if we want to check whether our graph values are reliable we can see from the code below
df.groupby(['Genre','Platform'])['NA_Sales'].mean()
```

```python
#distribution graph (histogram generally for int64)
sns.distplot(df.JP_Sales,bins=10,kde=False)
```
