**PYTHON DATA SCIENCE TOOLBOX**

```python
def tuble_ex():
    t = (1,2,3)
    return t
a,b,c=tuble_ex()
print(a,b,c)
```

```python
x=2
def f():
    x=3
    return x
print("global scope: ",x)
print("local scope : ",f())
```

```python
#nested func
def square():
    def add():
        x=2
        y=3
        z=x+y
        return z
    return add()**2
print(square())
```

```python
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4)

def f(**kwargs):
    for key,value in kwargs.items():
        print(key," : ",value)
f(countary="spain",capital="madrid",population=123456)
```

**LIST COMPREHENSION**
