# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'/kaggle/input/world-happiness/2019.csv')
```

# Dimension of data 

```python
df.shape
```

# checking for missing values

```python
df.isnull().sum()
```

# Distribution

```python
df.hist(bins=50,figsize=(15,15))
```

```python
df.head()
```

```python
df=df.rename(columns={'Country or region': 'Country'})
```

# Top 10 happiest countries

```python
df.groupby('Country')['Score'].max().sort_values(ascending=False).head(10).plot(kind='bar', figsize=(12,6),color='yellow')
plt.title('Top 10 happiest countries')
```

# 10 Saddest countries

```python
df.groupby('Country')['Score'].max().sort_values(ascending=False).tail(10).plot(kind='bar', figsize=(12,6),color='yellow')
plt.title('Top 10 Saddest countries')
```

# Top 20 countries with higghest GDP per capita

```python
df.groupby('Country')['GDP per capita'].max().sort_values(ascending=False).head(20).plot(kind='bar', figsize=(20,6),color='limegreen')
plt.title('Top 20 countries with higghest GDP per capita')
```

# 10 countries with lowest GDP per capita

```python
df.groupby('Country')['GDP per capita'].max().sort_values(ascending=False).tail(20).plot(kind='bar', figsize=(20,6),color='limegreen')
plt.title('20 countries with lowest GDP per capita')
```

# Top 20 countries with higghest social support

```python
df.groupby('Country')['Social support'].max().sort_values(ascending=False).head(20).plot(kind='bar', figsize=(20,6),color='cornflowerblue')
plt.title('Top 20 countries with higghest social support')
```

# 20 countries with lowest social support

```python
df.groupby('Country')['Social support'].max().sort_values(ascending=False).tail(20).plot(kind='bar', figsize=(20,6),color='cornflowerblue')
plt.title('20 countries with lowest social support')
```

# Top 20 countries with high healthy life expectancy

```python
df.groupby('Country')['Healthy life expectancy'].max().sort_values(ascending=False).head(20).plot(kind='bar', figsize=(20,6),color='aqua')
plt.title('Top 20 countries with high healthy life expectancy')
```

# 20 countries with lowest healthy life expectancy

```python
df.groupby('Country')['Healthy life expectancy'].max().sort_values(ascending=False).tail(20).plot(kind='bar', figsize=(20,6),color='aqua')
plt.title('20 countries with lowest healthy life expectancy')
```

# Top 20 countries with high freedom to make life choices

```python
df.groupby('Country')['Freedom to make life choices'].max().sort_values(ascending=False).head(20).plot(kind='bar', figsize=(20,6),color='plum')
plt.title('Top 20 countries with high freedom to make life choices')
```

# 20 countries with low freedom to make life choices

```python
df.groupby('Country')['Freedom to make life choices'].max().sort_values(ascending=False).tail(20).plot(kind='bar', figsize=(20,6),color='plum')
plt.title('20 countries with low freedom to make life choices')
```

# Top 20 countries with highest generosity

```python
df.groupby('Country')['Generosity'].max().sort_values(ascending=False).head(20).plot(kind='bar', figsize=(20,6),color='red')
plt.title('Top 20 countries with highest generosity')
```

# 20 countries with lowest generosity

```python
df.groupby('Country')['Generosity'].max().sort_values(ascending=False).tail(20).plot(kind='bar', figsize=(20,6),color='red')
plt.title('20 countries with lowest generosity')
```

# Top 20 countries with highest perceptions of corruption

```python
df.groupby('Country')['Perceptions of corruption'].max().sort_values(ascending=False).head(20).plot(kind='bar', figsize=(20,6),color='lime')
plt.title('Top 20 countries with highest perceptions of corruption')
```

# 20 countries with lowest perceptions of corruption

```python
df.groupby('Country')['Perceptions of corruption'].max().sort_values(ascending=False).tail(20).plot(kind='bar', figsize=(20,6),color='lime')
plt.title('20 countries with lowest perceptions of corruption')
```

# What makes the people happy

```python
df.head()
```

```python
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(12, 6),
                         sharey='row', sharex='row') #we created a grid of 6 plots

axes[0, 0].scatter(x=df['Social support'],y=df['Score']) # plot at 1st row, 1st col and 1st position
axes[0, 0].set_xlabel('Social support')
axes[0, 0].set_ylabel('Score')

axes[0, 1].scatter(x=df['Healthy life expectancy'],y=df['Score'])# plot at 1st row, 2st col and 2st position
axes[0, 1].set_xlabel('Healthy life expectancy')
axes[0, 1].set_ylabel('Score')

axes[1, 0].scatter(x=df['Freedom to make life choices'],y=df['Score'])# plot at 2nd row, 1st col and 1st position
axes[1, 0].set_xlabel('Freedom to make life choices')
axes[1, 0].set_ylabel('Score')

axes[1, 1].scatter(x=df['Generosity'],y=df['Score'])# plot at 2nd row, 2nd col and 2nd position
axes[1, 1].set_xlabel('Generosity')
axes[1, 1].set_ylabel('Score')

axes[2, 0].scatter(x=df['Perceptions of corruption'],y=df['Score'])# plot at 3rd row, 1st col and 1st position
axes[2, 0].set_xlabel('Perceptions of corruption')
axes[2, 0].set_ylabel('Score')

axes[2, 1].scatter(x=df['GDP per capita'],y=df['Score'])# plot at 3rd row, 2nd col and 2nd position
axes[2, 1].set_xlabel('GDP per capita')
axes[2, 1].set_ylabel('Score')




fig.tight_layout()

```

## from the above fig we can see that, except perception of corruption every other thing like social, support, health life expectancy and so on makes people happy.
