# Importing libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'/kaggle/input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
```

```python
df.shape#dimension
```

```python
df.head()
```

```python
#count of fiction, non fiction books 
df.groupby('Genre').count()
```

# Maximum user rating for fiction and non fiction

```python
#max user ratings for diffrent genre
df.groupby('Genre')['User Rating'].max().plot(kind='bar',figsize=(10,6),color='slategrey')
plt.title('Max user rating for different genre')
plt.xlabel('Genre')
plt.ylabel('User Rating')
plt.show()
```

We can see that for fiction and non fiction max rating is same

# Minimum user rating for fiction and non fiction

```python
#min user ratings for diffrent genre
df.groupby('Genre')['User Rating'].min().plot(kind='bar',figsize=(10,6),color='lightsteelblue')
plt.title('Min user rating for different genre')
plt.xlabel('Genre')
plt.ylabel('User Rating')
plt.show()
```

We can see that for fiction the min rating is around 3.5 and for non fiction min rating is 4

# Top 10 authors with highest rated books

```python
df.groupby('Author')['User Rating'].max().sort_values(ascending=False).head(10).plot(kind='bar',figsize=(12,6),color='cornflowerblue')
plt.title('Top 10 authors with highest ratings')
```

# Top 10 Books with highest ratings

```python
df.groupby('Name')['User Rating'].max().sort_values(ascending=False).head(10).plot(kind='bar',figsize=(12,6),color='royalblue')
plt.title('Top 10 Books with highest ratings')
```

# 10 books with highest reviews

```python
df.groupby('Name')['Reviews'].max().sort_values(ascending=False).head(10).plot(kind='bar',color="blue",figsize=(12,6))
plt.xlabel('Books')
plt.ylabel('No of Reviews')
plt.title('10 books with highest reviews')
```

# Heatmap

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True)
```

# Percentage of fiction and non fiction books

```python
color=['teal', 'cyan']
df['Genre'].value_counts().plot(kind='pie',figsize=(8,8),autopct="%1.2f",colors=color,explode=[0,0.02])
plt.axis=('equal')
plt.legend(loc='upper right')
plt.show()
```
