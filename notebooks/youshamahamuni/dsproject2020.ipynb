As seen in the output above, the columns company,agent,country, and children have NaN or NA values. Using the output below, we can see the type of the data in each column. This information can be used to decide what to impute the NaN values with. However, before that, it can be seen that the total number of observations are 119,000 approximately, and the number of NaN values in the company column are 112,593 which means a large majority of the values in the column are null. We can then go on to exclude that column from the data we will be analyzing. 

```python
data.dtypes
```

```python
del data['company']
```

Now, we can analyze the other columns and impute their NaN values. Let us check the column agent. It has a data type of float64.

```python
data['agent']
```

As seen in the preview above, agent column has data type float. In order to fill the NaN values, we can use the mean. Let us now compute the mean. 

```python
data['agent'].mode()
```

```python
data['agent'].fillna(9.0,inplace=True)
```

We have now replaced the NaN values with the mean. Now let us move on to checking the NaN values in the country column. Country has data of object type.

```python
data['country']
```

Since the country variable is an object type, it can be categorised as categorical data. For categorical data, the best possible way to fill the NaN cells, are by imputing the mode of the variable. Let us now find the mode for the country column.

```python
data['country'].mode()
```

```python
data['country'].fillna('PRT',inplace=True)
```

```python
data['children']
```

```python
data['children'].mode()
```

```python
data['children'].fillna(0,inplace=True)
```

```python
data['children']
```

```python
data['children']=data['children'].apply(np.int64)
```

```python
Q1=data.quantile(.25)
Q3=data.quantile(.75)
IQR=Q3-Q1
```

```python
outlier = data[((data['total_of_special_requests'] < (Q1['total_of_special_requests'] - 1.5 * IQR['total_of_special_requests'])) | (data['total_of_special_requests'] > (Q3['total_of_special_requests'] + 1.5 * IQR['total_of_special_requests'])))].index
data.drop(outlier, inplace= True)
```

```python
data.boxplot(column=['total_of_special_requests'])
data['total_of_special_requests'].describe()
```

```python
outlier2 = data[((data['required_car_parking_spaces'] < (Q1['required_car_parking_spaces'] - 1.5 * IQR['required_car_parking_spaces'])) | (data['required_car_parking_spaces'] > (Q3['required_car_parking_spaces'] + 1.5 * IQR['required_car_parking_spaces'])))].index
data.drop(outlier2, inplace= True)
```

```python
data.boxplot(column=['required_car_parking_spaces'])
data['required_car_parking_spaces'].describe()
```

```python
df_us = data[data.country == 'USA']
df_uk = data[data.country == 'GBR']
df_po = data[data.country == 'PRT']
df_ger = data[data.country == 'DEU']
df_sp = data[data.country == 'ESP']
df_merged = pd.concat([df_us, df_uk, df_po, df_ger, df_sp])
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(x= 'stays_in_weekend_nights', y='stays_in_week_nights', hue = 'country', data = df_merged )
print("scatter plot depicts Stay_in_week_nights vs. Stay_in_weekend_nights in the five countries")
```

```python
plt.figure(figsize=(15,5))
sns.scatterplot(x= 'country', y='previous_cancellations', hue = 'country', data = df_merged )
print("scatter plot depicts previous_cancellations vs. country in the five countries")
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x= 'arrival_date_month', y = 'booking_changes', data = df_us)
print("chart shows in which month US visitors are arriving the most & also the booking change frequency")
```

```python

plt.figure(figsize=(10,10))
sns.countplot(x='hotel', data = df_us, palette='gist_earth')
plt.title('Hotel Types', weight='bold')
plt.xlabel('Hotel', fontsize=12)
plt.ylabel('Count', fontsize=12)
```
