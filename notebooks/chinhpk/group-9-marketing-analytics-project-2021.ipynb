# **4) Data Preperation**

```python
airline = airline.drop(columns=["id", "Gender", "Age"])
```

```python
airline.head()
```

Ideas of new features:

* Total delay = Departure + Arrival Delay

* Overall customer satisfaction = all

* Pre-flight experience = Ease of Online booking, Online boarding, Departure/Arrivel time convenient

* Airport experience = Checkin service, Baggage handling, Gate location

* Inflight experience = Cleanliness, Inflight service, Leg room service, Inflight entertainment, Seat comfort, Food and drink, On-board service

<img src="https://www.mycustomer.com/sites/default/files/green3_1.jpg" width="800">

Source: https://www.mycustomer.com/experience/engagement/customer-journey-mapping-vs-process-design-do-you-know-the-difference


```python
# Creation of the new feature "total delay"
airline["total_delay"] = airline["Departure Delay in Minutes"] + airline["Arrival Delay in Minutes"]
# Creation of the new feature "Pre-flight experience"
airline["pre-flight experience"]= round((airline["Ease of Online booking"] + airline["Online boarding"] + airline["Departure/Arrival time convenient"] )/3, 2)
# Creation of the new feature "Airport experience"
airline["airport experience"]= round((airline["Checkin service"] + airline["Baggage handling"] + airline["Gate location"] )/3, 2)
# Creation of the new feature "Inflight experience"
airline["inflight experience"]= round((airline["Cleanliness"] + airline["Inflight service"] + airline["Leg room service"] + airline["Inflight entertainment"] + airline["Seat comfort"] + airline["Food and drink"] + airline["On-board service"])/7, 2)
```
