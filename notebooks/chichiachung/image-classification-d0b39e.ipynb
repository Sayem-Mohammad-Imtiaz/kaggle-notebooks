```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('../input/555555/fullsizeoutput_189.jpeg')
img = img.resize((224,224))
plt.imshow(img)
```

```python
import numpy as np
test_x = np.array(img) / 255.0
print(test_x.shape)
```

```python
test_x = test_x.reshape(1,224,224,3)
```

## Load Model

```python
# Import Model
#from tensorflow.keras.applications import VGG16
#from tensorflow.keras.applications import ResNet101V2
from tensorflow.keras.applications import InceptionV3

#from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
#from tensorflow.keras.applications.resnet import preprocess_input, decode_predictions
from tensorflow.keras.applications.inception_v3 import preprocess_input, decode_predictions

# Load Model
#model = VGG16(weights='imagenet')
#model = ResNet101V2(weights='imagenet')
model = InceptionV3(weights='imagenet')
```

## Prediction

```python
# model prediction
preds = model.predict(test_x)
# decode prediction
dec_preds =  decode_predictions(preds, top=3)[0]
print('Predicted:', dec_preds)
```
