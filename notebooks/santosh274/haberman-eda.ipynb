EDA on Haberman data for cancer patient survival rate

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
haberman=pd.read_csv("../input/haberman.csv")
haberman.head()
```

```python
#Adding column name to dataset
columns=['Age','OperationYear','AuxillaryNode','Survival']
haberman_data=pd.read_csv('../input/haberman.csv',names=columns)
haberman_data.head()
```

```python
haberman_data.shape
```

```python
haberman_data.Survival.value_counts()
```

```python

sns.FacetGrid(haberman_data, hue="Survival", size=7) \
   .map(sns.distplot, "Age") \
   .add_legend()
plt.show()
```

```python

sns.FacetGrid(haberman_data, hue="Survival", size=7) \
   .map(sns.distplot, "OperationYear") \
   .add_legend()
plt.show()
```

```python

sns.FacetGrid(haberman_data, hue="Survival", size=7) \
   .map(sns.distplot, "AuxillaryNode") \
   .add_legend()
plt.show()
```

```python
less_than_five = haberman_data.loc[haberman_data["Survival"]==2]
more_than_five = haberman_data.loc[haberman_data["Survival"]==1]
```

```python


plt.subplot(1, 2, 1)
counts, bin_edges = np.histogram(less_than_five['Age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(['Pdf  Survived',
            'Cdf  Survived'])

plt.subplot(1, 2, 2)
counts, bin_edges = np.histogram(more_than_five['Age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(['Pdf Died',
            'Cdf Died'])

plt.show()

```

```python


plt.subplot(1, 2, 1)
counts, bin_edges = np.histogram(less_than_five['OperationYear'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(['Pdf  Survived',
            'Cdf  Survived'])

plt.subplot(1, 2, 2)
counts, bin_edges = np.histogram(more_than_five['OperationYear'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend(['Pdf Died',
            'Cdf Died'])

plt.show()

```

Observations:
            1.Taking OperationYear as parameter ploting CDF and PDF curves, we observe both classes have over
              lapping each other
            2. OperationYear cannot be taken as parameter for analysis
