FINDING TOTAL COST AND ADDING IT AS A COLOUMN TO EXISTING DATASET

```python
a=[]
for i in range (0,len(df)):
  b=df['Quantity'][i]*df['UnitPrice'][i]
  a.append(b)
df['TotalCost'] = np.array(a)
```

```python
df.head()
```

DROPPING DUPLICATES

```python
df=df.drop_duplicates()
len(df)
```

FINDING THE TOTALCOST SPENT PER TRANSACTION

```python
mt1=df.groupby(['InvoiceNo'])['TotalCost'].sum()
mt1
```

MAXIMUM AMOUNT SPENT PER TRANSACTION

```python
max_cost=sorted(mt1 ,reverse=True)

print(max_cost[0:1])
```

FINDING THE AVERAGE AMOUNT SPENT PER TRANSACTION

```python
mt=df.groupby(['InvoiceNo'])['TotalCost'].mean()
mt
```

DATA VISUALISATION

Frequency of most bought items itemcode wise

```python
plt.rcParams['figure.figsize'] = (18, 7)
color = plt.cm.copper(np.linspace(0, 1, 40))
df['StockCode'].value_counts().head(40).plot.bar(color = color)
plt.title('frequency of most bought items itemcode wise', fontsize = 20)
plt.xticks(rotation = 90 )
plt.grid()
plt.show()
```

Tree map for most 50 popular items

```python
y = df['Description'].value_counts().head(50).to_frame()
y.index
```

```python
plt.rcParams['figure.figsize'] = (30, 30)
color = plt.cm.cool(np.linspace(0, 1, 50))
squarify.plot(sizes = y.values, label = y.index, alpha=.8, color = color)
plt.title('Tree Map for Popular Items')
plt.axis('off')
plt.show()
```
