```python
import numpy as np
import pandas as pd
import random

from sklearn.metrics import classification_report, log_loss, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import ShuffleSplit

from xgboost import XGBClassifier
```

# Data Process

```python
data = pd.read_csv("../input/heart-disease-uci/heart.csv")
data[0:2]
```

```python
n=len(data)
print(n)
```

```python
N=[]
for i in range(n):
    N+=[i]
random.shuffle(N)
```

```python
datay=data['target']
data2=data.drop('target',axis=1)
```

```python
data2=np.array(data2)
datay=np.array(datay)
```

```python
train=data2[N[0:(n//10)*8]]
test=data2[N[(n//10)*8:]]
trainy=datay[N[0:(n//10)*8]]
testy=datay[N[(n//10)*8:]]
```

```python
X = train
y = trainy
```

# Modeling

```python
clf = XGBClassifier(max_depth=3, n_estimators=1000, learning_rate=0.01)
```

```python
ss = ShuffleSplit(n_splits=5,train_size=0.8,test_size =0.2,random_state=0) 

for train_index, test_index in ss.split(X): 
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = y[train_index], y[test_index]
    clf.fit(X_train, Y_train) 
    print(clf.score(X_test, Y_test))
```

# Predict

```python
PRED = clf.predict(test)
PRED[0:20]
```

```python
ANS=testy
ANS[0:20]
```

```python
accuracy=accuracy_score(ANS,PRED)
print(accuracy)
```
