```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

import bar_chart_race as bcr
from IPython.display import HTML
import warnings
warnings.filterwarnings("ignore")
from IPython.display import Video
```

# Data preparation

```python
data_dir='../input/world-happiness-report'
```

```python
paths=[]
for i in range(6):
    item=str(2015+i)+'.csv'
    paths+=[os.path.join(data_dir,item)]
paths
```

```python
data0=pd.read_csv(paths[5])
print(data0.columns.tolist())
data0
```

```python
data1=data0.drop(['Regional indicator'],axis=1)
data1
```

# Dataframe for Bar Chart Race

```python
# set double indexes
data2=data1.set_index('Country name').T
data2.index.name="Item"
data2
```

```python
data4=data2.fillna(0)
```

# Slide Show using Bar Chart Race

```python
df = data4
bcr.bar_chart_race(
    df=df,
    filename='WorldHappiness_BarChartRace.mp4',
    orientation='h',
    sort='desc',
    n_bars=20,
    fixed_order=False,
    fixed_max=False,
    steps_per_period=1,  # 1:slideshow
    interpolate_period=False,
    label_bars=False,
    bar_size=.85,   
    period_label={'x': .99, 'y': .05, 'ha': 'right', 'va': 'center'},
    #period_fmt='%B %d, %Y',
    period_summary_func=lambda v, r: {'x': .99, 'y': .10,'s': '',
                                      'ha': 'right', 'size': 8, 'family': 'Courier New'},
    perpendicular_bar_func='median',
    period_length=1500,
    figsize=(5,3),
    dpi=500,
    cmap='dark24',
    title='World Happiness Ranking 2015-2020',
    title_size=15,
    bar_label_size=7,
    tick_label_size=5,
    shared_fontdict={'color' : '.1'},
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .7},
    filter_column_colors=True)
```
