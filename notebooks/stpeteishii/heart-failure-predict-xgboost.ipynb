```python
import numpy as np
import pandas as pd
import category_encoders as ce

from sklearn.metrics import classification_report, log_loss, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import ShuffleSplit

from xgboost import XGBClassifier
```

```python
data = pd.read_csv("../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv")
data[0:2].T
```

```python
n=len(data)
print(n)
```

```python
train=data[0:(n//10)*8]
y=train['DEATH_EVENT']
train2=train.drop('DEATH_EVENT',axis=1)
print(y.value_counts())
```

```python
test=data[(n//10)*8:]
testy=test['DEATH_EVENT']
test2=test.drop('DEATH_EVENT',axis=1)
print(testy.value_counts())
```

```python
data.info()
```

```python
X_train0 = train2
X_test0 = test2
```

```python
X = np.array(X_train0)
y = np.array(y)
```

```python
clf = XGBClassifier(max_depth=3, n_estimators=1000, learning_rate=0.01)
```

```python
ss = ShuffleSplit(n_splits=5,train_size=0.8,test_size =0.2,random_state=0) 

for train_index, test_index in ss.split(X): 
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = y[train_index], y[test_index]
    clf.fit(X_train, Y_train) 
    print(clf.score(X_test, Y_test))
```

```python
pred0 = clf.predict(np.array(X_test0))
```

```python
PRED=[]
for item in pred0:      
    PRED+=[item]
PRED[0:20]
```

```python
ANS=list(testy)
ANS[0:20]
```

```python
accuracy=accuracy_score(ANS,PRED)
print(accuracy)
```
