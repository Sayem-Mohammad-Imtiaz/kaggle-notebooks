```python
list_cols = ['MouseID','Genotype','Treatment','Behavior','class']
ce_ohe = ce.OrdinalEncoder(cols=list_cols,handle_unknown='impute')
data2 = ce_ohe.fit_transform(data)
data2.head()
```

```python
data2.isnull().sum()
```

```python
data2=data2.dropna()
data2.isnull().sum()
```

```python
n=len(data2)
print(n)
```

```python
train = data2[0:(n//10)*8]
test = data2[(n//10)*8:]
```

```python
y_train0 = train['class']
y_test0 = test['class']
X_train0 = train.drop('class', axis = 1)
X_test0 = test.drop('class', axis = 1)
```

```python
X = np.array(X_train0)
y = np.array(y_train0)
```

```python
clf = XGBClassifier(max_depth=3, n_estimators=1000, learning_rate=0.01)
```

```python
ss = ShuffleSplit(n_splits=5,train_size=0.8,test_size =0.2,random_state=0) 

for train_index, test_index in ss.split(X): 

    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = y[train_index], y[test_index]

    clf.fit(X_train, Y_train) 
    print(clf.score(X_test, Y_test))
```

```python
pred0 = clf.predict(np.array(X_test0))
```

```python
PRED=pred0
ANS=y_test0
accuracy=accuracy_score(ANS,PRED)
print(accuracy)
```
