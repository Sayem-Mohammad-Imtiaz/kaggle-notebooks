# Applying the ML Algorithms

```python
# Importing libraries for evaluation
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

## Splitting the data into test and train

```python
from sklearn.model_selection import train_test_split
X=dataset.drop('diagnosis',axis=1,inplace=False)
y=dataset['diagnosis']
print(X.head(2))
print(y.head(2))
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=455)
```

##  Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lm=LogisticRegression()
lm.fit(X_train,y_train)
y_pred=lm.predict(X_test)
# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
# Accuracy score
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,y_test))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
#Using cross validation for KNN
X_t,X_test,y_t,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
X_train,X_cv,y_train,y_cv=train_test_split(X_t,y_t,test_size=0.3,random_state=42)
```

```python
error_rate=[]

for i in range(1,100):
    KNN=KNeighborsClassifier(n_neighbors=i)
    KNN.fit(X_train,y_train)
    pred_i=KNN.predict(X_cv)
    error_rate.append(np.mean(pred_i != y_cv))
```

```python
plt.figure(figsize=(20,12))
plt.plot(range(1,100),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
```

Choosing the N with lowest error rate i.e 5 or 7

```python
model=KNeighborsClassifier(n_neighbors=5)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
# Summary of the predictions made by the classifier
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
# Accuracy score
from sklearn.metrics import accuracy_score
print('accuracy is',accuracy_score(y_pred,y_test))
```
