13. Data Visualisation

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig = plt.gcf()
fig.set_size_inches(8, 5)
cplot =sns.countplot(y="service", data=df)
cplot.set_title("repartition of services")
plt.show()
plt.figure(figsize=(30, 15)),
cplot2 = sns.countplot(y="proto", data=df)
cplot2.set_title("Repartition of protocols")
plt.show()
cplot3 = sns.countplot(y="attack_cat", data=df)
cplot3.set_title("Repartition of attack")
plt.show()

```

14 Correlation Matrix

```python
plt.figure(figsize=(30, 15))
matrix = sns.heatmap(df.corr(), vmin=-1, vmax=1, annot=True, cmap="BrBG")
matrix.set_title('Correlatio matrix', fontdict={'fontsize':12}, pad=10)
plt.show
```

First of all, we note a lack of information about the services, the majority (40,000 lines) of the services are not identified. The use of the DNS service is the most important with more than 20 000 lines.  As far as the protocols are concerned, the majority are represented by UDP and TCP. We also notice that about 50% of the traffic is normal and for the rest of the attacks are rather "Generic" or "Exploits". 
To interpret the correlation matrix we need to focus on the darkest cells off the diagonal.  A strong positive correlation is found between several variables, this is represented by dark green boxes. Other variables are negatively correlated, represented by dark orange boxes.
In order to deepen the analysis on traffic, we can start to discard the variables that are not correlated with the "label" variable or study the interdependence between the variables with the objective of excluding the most dependent variables.

It is therefore necessary to make the collection of flows more reliable, especially for the service variable where the identification of data is missing. I also recommend using the SSH protocol to encrypt the flows.
