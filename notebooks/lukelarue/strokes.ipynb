The distribution of average glucose levels is interesting, it appears the population in the dataset has two means, one centered around 80 and the other centered around 220. After some very brief googling, I believe there are two possible explanations for this. The first explanation is there is one population with diabetes, and thus higher levels of glucose than those without diabetes. Another explanation is that people have different glucose levels depending on the time of day and how recently they have eaten, so those measured right before a meal would have lower glucose levels, and those measured after a meal have higher levels. However, based off a quick google search, it appears that those without diabetes do not have great variation of their glucose levels before and after meals, whereas those with diabetes do. So it appears a signifcant number of individuals in the dataset have diabetes, thus creating the second peak.

```python
sns.displot(df.avg_glucose_level, kde = False);
plt.title('Distribution of Average Glucose Level');
```
