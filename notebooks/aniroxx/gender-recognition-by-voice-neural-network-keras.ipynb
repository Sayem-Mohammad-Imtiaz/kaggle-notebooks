```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.metrics as metrics

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical

import keras
from keras.layers import Dense
from keras.models import Sequential
from keras.callbacks import History 
from keras.utils import plot_model
from keras.optimizers import SGD
```

```python
data = pd.read_csv('/kaggle/input/voicegender/voice.csv')
```

```python
"""Model Building"""
#Split features and labels
x = data.iloc[:, :-1].values
y = data.iloc[:,-1].values
```

```python
gender_encoder = LabelEncoder()
y = gender_encoder.fit_transform(y) #Male = 1 ----- Female = 0 
y
```

```python
#Preprocessess data
scaler = StandardScaler()
x = scaler.fit_transform(x)
x
```

```python
#Test-train split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .1)
```

```python
#Build Neural Network 

n_cols = x_train.shape[1]
y_train = to_categorical(y_train, 2)

hist = History()

model = Sequential()

model.add(Dense(1000, activation='relu', input_dim = n_cols))
model.add(Dense(1000, activation='relu'))
model.add(Dense(1000, activation='relu'))
model.add(Dense(1000, activation='relu'))


model.add(Dense(2, activation='softmax'))

model.compile(optimizer='adam', 
              loss='binary_crossentropy', 
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs = 40, validation_split = .1, callbacks = [hist])
```

```python
y_pred =  model.predict(x_test)
y_pred = np.round(y_pred[:,1])
print(metrics.accuracy_score(y_pred,y_test))
```
