# Easy peasy lemon squeezy regression, just for fun.

### Easy peasy dataset has been picked, we continue.

### First, let's see what's in there...

```python
train = pd.read_csv('/kaggle/input/house-price-prediction-challenge/train.csv')
train.head()
```

### Let's drop a few that may cause us trouble and see where we get with it.

```python
X = train.copy()
Y = X['TARGET(PRICE_IN_LACS)']
X.drop(['TARGET(PRICE_IN_LACS)'], axis = 1, inplace=True)
X.drop(['BHK_OR_RK', 'ADDRESS', 'POSTED_BY'], axis=1, inplace=True)
X.isnull().values.any()
```

### Now, let's see if there are any correlations in the data. 

```python
correlation_df = train.corr()
sns.set(rc={'figure.figsize':(15,15)})
sns.heatmap(correlation_df)#annot=True
plt.show()
```

### Seems to be. Let's see if we can find something linear.

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X, Y)
print(reg.score(X, Y))
```

### ... That doesn't sound promising. let's cheat and do something nonlinear.

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor().fit(X,Y)
print(dtr.score(X, Y))
```

### Wasn't linear. :|

### Now, let's just predict the test set and clean up.

```python
test = pd.read_csv('/kaggle/input/house-price-prediction-challenge/test.csv')
test.head()
```

```python
test.drop(['BHK_OR_RK', 'ADDRESS', 'POSTED_BY'], axis=1, inplace=True)
res = dtr.predict(test)
```

```python
sample = pd.read_csv('/kaggle/input/house-price-prediction-challenge/sample_submission.csv')
sample.head()
```

```python
for i in range(0, res.shape[0]):
    sample.iloc[i,0] = res[i]
```
