# Prediction of Breast Cancaer Sub-types according to Proteome data

### Rationale
There are 4 different breast cancer sub-types: 'basal-lie', 'HER2', 'LumA', 'LumB'. **The goal here is to predict breast cancer sub-types according to proteome data and advance the prediction precision** 

One key goal is to **extracting the correct feature sets (protein sets in this case)** for the prediction. It is a combination of application of domain knowledge and statistical analysis. In the breast_cancer_proteosome_analysis, I directly chose the PAM50 genes as the feature sets. Here, I will explore the 'statistical way' to extract features using scikit learn.

notes: **sklearn.feature_selection**. Notice here I didn't try feature union here, which is pretty handy most of the time. (I tried, the performance is not getting better.)

Another key goal is to finetune a ML pipeline to get better performance. I've already explored this datasets with 2 machine learning models: K-neighbors and svm.SVC. However, I didn't explore all the way to finetune the models to get better performance. 

The third key point is utilizing **pipeline**. Pipelines work by allowing for a linear sequence of data transforms to be chained together culminating in a modeling process that can be evaluated. The goal is to ensure that all of the steps in the pipeline are constrained to the data available for the evaluation. **This is very important.**
