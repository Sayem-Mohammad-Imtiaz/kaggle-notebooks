# Gesture recognition using EMG
## Dataset
 1. Sensor readings of dimension 64 would predict a gesture class (0-3).
    Gesture classes were : rock - 0, scissors - 1, paper - 2, ok - 3.  
    Structure of data : 
    [8sensors][8sensors][8sensors][8sensors][8sensors][8sensors][8sensors][8sensors][GESTURE_CLASS]

```python
data0 = pd.read_csv("/kaggle/input/emg-4/0.csv", header=None) # For rock[0]
data1 = pd.read_csv("/kaggle/input/emg-4/1.csv", header=None) # For scissors[1]
data2 = pd.read_csv("/kaggle/input/emg-4/2.csv", header=None) # For paper[2]
data3 = pd.read_csv("/kaggle/input/emg-4/3.csv", header=None) # For ok[3]
data = pd.concat([data0,data1,data2,data3], axis=0)
data.iloc[:5]
```

```python
from sklearn.model_selection import train_test_split
Y = data.iloc[:,64]
X = data.drop(data.columns[64], axis=1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8, random_state=1, stratify=Y)
```

```python
print(X_train.shape)
print(X_test.shape)
print(np.mean(X_train, axis=0))
```

### Data normalization

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = pd.DataFrame(sc.fit_transform(X_train))
X_test = pd.DataFrame(sc.transform(X_test))
```

### Classification

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, accuracy_score, f1_score
model = SVC(kernel = 'rbf')
model.fit(X_train, Y_train)
predictions = model.predict(X_test)
print(accuracy_score(Y_test, predictions))
print(classification_report(Y_test, predictions))
```
