```python
epl = pd.read_csv('/kaggle/input/formula-1-world-championship-1950-2020/races.csv')
epl.head()
```

#read the Patientinfo, Time, and Region csv fileinto a dataframe


```python
# Number of rows and columns
rows, cols = epl.shape
print('Number of race: {}'.format(rows))
print('Number of stats per race: {}'.format(cols))
```

```python
list(epl.columns)
```

```python
epl.info()
```

```python
epl_name_time=epl.groupby('name').time.value_counts()
epl_name_time
```

```python
epl.describe()
```

```python
# Number of race in f1 by each time
epl_time = epl['time'].value_counts()
print(epl_time)
```

**visualization**

```python
fig = px.pie(epl_time, values=epl_time.values, names=epl_time.index)
fig.show()
```

```python
pd.crosstab(epl['name'], epl['time']).plot(kind='bar',figsize=(15,15))
```

**Prix by name of the Formula 1**

```python
plt.figure(figsize= (14,8))

ax = sns.countplot(x='name', data=epl)
ax.set_title(label='Count of names of the race', fontsize=10)

ax.set_xlabel(xlabel='name')
ax.set_ylabel(ylabel='Count')

plt.show()
```

```python
print("Faster prix/s: ")
epl.loc[epl['name'] == epl['name'].max()]
```

```python
print("Lowest prix/s: ")
epl.loc[epl['name'] == epl['name'].min()]
```
