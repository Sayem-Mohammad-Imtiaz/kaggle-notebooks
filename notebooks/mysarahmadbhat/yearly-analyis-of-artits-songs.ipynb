## AS we can see some of the data is missing from last_week column

## Let's Handle that First
## Impute the missing values

```python
df["last-week"]=df["last-week"].fillna(df["last-week"]).median()
```

## So as we can see we clean the data .
## There is no null values present in the data set

## Let's do some analysis

## 1st change the date column in Time stamp
   - so that we can make a new columns YEAR & Month

```python
df["date"]=pd.DatetimeIndex(df["date"])
```

```python
df["year"]=pd.DatetimeIndex(df["date"]).year
```

```python
df["month"]=pd.DatetimeIndex(df["date"]).month
```

## 01.Top 10 Featured songs of all time

```python
x=df.groupby(by=df["song"])["year"].value_counts().sort_values(ascending=False)
x.head(10).plot(kind="bar")
```

## 02. Top 10  Featured Artist Of All Time

```python
x=df.groupby(by=df["artist"])["year"].value_counts().sort_values(ascending=False)
x.head(10).plot(kind="bar")
```
