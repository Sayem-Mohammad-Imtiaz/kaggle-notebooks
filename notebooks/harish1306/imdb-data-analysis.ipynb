***Let's seperate each of the data***

```python
title = pd.read_csv("/kaggle/input/imdb-extensive-dataset/IMDb title_principals.csv")
```

```python
names = pd.read_csv("/kaggle/input/imdb-extensive-dataset/IMDb names.csv")
```

```python
movies = pd.read_csv("/kaggle/input/imdb-extensive-dataset/IMDb movies.csv")
```

```python
ratings = pd.read_csv("/kaggle/input/imdb-extensive-dataset/IMDb ratings.csv")
```

## Joining the data based on our wish I am plannig to see the most popular movie

```python
output1 = pd.merge(title,ratings, 
                   on='imdb_title_id', 
                   how='inner')
```

```python
output1.head()
```

```python
output2 = pd.merge(output1,movies,
                  on = "imdb_title_id",
                  how="inner")
```

```python
output2.head()
```

```python
output2.columns
```

***Ahhh!! There are enoromous no of columns***

**Let's prepare the dataframe with columns we want**

```python
df = pd.DataFrame(output2 , columns = ['imdb_title_id', 'total_votes','title', 'original_title', 'year', 'genre',
       'duration', 'country', 'language','budget', 'worlwide_gross_income','metascore'])
```

```python
df.head()
```

**Let's see the Top 25 movies**

```python
viz = df.sort_values(by=['total_votes',"year"], ascending=False).head(25)
```

```python
viz.head()
```

```python
viz['worlwide_gross_income'] = viz['worlwide_gross_income'].str.replace('$', '')
```

```python
viz.dtypes
```

```python
viz["worlwide_gross_income"] = viz["worlwide_gross_income"].astype(int)
viz.dtypes
```

```python
viz.head()
```

```python
import matplotlib.pyplot as plt
```

**Let's play with vizualization**

```python
viz.plot(x="original_title", y="total_votes", kind="bar")
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = plt.axes(projection ='3d')

z = viz["worlwide_gross_income"]
x = viz["metascore"]
y = viz["total_votes"]
fig.set_figwidth(40)
fig.set_figheight(10)

ax.plot_trisurf(x,y,z ,linewidth=0, antialiased=False)
ax.set_title('Reletion between metascore, votes , income')
plt.show()
```
