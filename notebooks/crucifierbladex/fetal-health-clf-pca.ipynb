```python
from sklearn.model_selection import train_test_split
x=df.drop(['fetal_health'],axis=1)
y=df['fetal_health'].values
```

```python
x_train,y_train,x_test,y_test=train_test_split(x,y,test_size=0.1,random_state=123)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train,y_train=scaler.fit_transform(x_train),scaler.fit_transform(y_train)
```

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(x_train,x_test)
```

```python
xgb.score(x_train,x_test)
```

```python
xgb.score(y_train,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
ada=AdaBoostClassifier()
rf=RandomForestClassifier()
dt=DecisionTreeClassifier()
```
