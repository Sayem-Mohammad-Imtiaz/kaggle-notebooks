```python
data=pd.read_csv('../input/chess/games.csv')
```

```python
threshold = sum(data.turns) / len(data.turns)
print(threshold)
```

```python
data["turns_level"]= [ "high" if i > threshold else "low" for i in data.turns]
print(data["turns_level"])
data.loc[:20,["turns_level" , "turns"]]
```
