***Visualizing the features and response***

```python
for col in df_tr.columns:
    if df_tr[col].dtype=='int64':
        plt.hist(df_tr[col].value_counts(),Label=True)
        plt.title('Histogram for {}'.format(col))
        plt.show(); 
    
    else:
        plt.pie(df_tr[col].value_counts(),labels=df_tr[col].unique())
        plt.title('Piechart for {}'.format(col))
        plt.show();
```

**Visualizing the relation**

```python
for col in df_tr.columns:
    if df_tr[col].dtype != 'object':
        sns.catplot(x=col, y='y',kind="boxen",data=df_tr.sort_values(col))
        plt.title('Catplot for {}'.format(col))
        plt.show();
```

***Creatibg Dummies for categorical columns***

```python
dftr=pd.get_dummies(df_tr,prefix=None,prefix_sep='_',drop_first=True)
```

```python
dfte=pd.get_dummies(df_te,prefix=None,prefix_sep='_',drop_first=True)
```

```python
dftr
```

```python
dftr.info()
```

***Differntiating X,y***

```python
xtr=dftr.iloc[:,:42]
ytr=dftr.iloc[:,42]
```

```python
xte=dfte.iloc[:,:42]
yte=dfte.iloc[:,42]
```

**Transforming X's**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(xtr)
X_train = scaler.transform(xtr)
X_test = scaler.transform(xte)
```

***KNN***

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 2)
classifier.fit(X_train, ytr)
```

```python
y_pred = classifier.predict(X_test)
y_pred
```

***Visualizing Our accuracy***

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
result = confusion_matrix(yte, y_pred)
print("Confusion Matrix:")
print(result)
result1 = classification_report(yte, y_pred)
print("Classification Report:",)
print (result1)
result2 = accuracy_score(yte,y_pred)
print("Accuracy:",result2)
```
