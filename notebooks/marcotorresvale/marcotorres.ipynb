## 5.Comparacion entre Arbol de decision ,Random forest y Gradient Boosting

```python
## se importan librerias del arbol de decision cargando datos de entrenamiento y test o validacion
from sklearn.tree import DecisionTreeRegressor
treeD = DecisionTreeRegressor(max_depth=None)

treeD.fit(x_train, y_train)

tree_score = treeD.score(x_validation, y_validation)
print('Tree Score: ', tree_score)
```

## 7.Neural Networks

```python
## se importan librerias de keras y tensor flow
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.callbacks import EarlyStopping
%matplotlib inline
%config InlineBackend.figure_format='retina'
```

```python
##se caragan variables de entrenamiento y test
x_train1, x_test1, y_train1, y_test1 = train_test_split(Auto_fs, y_fs, test_size=0.2,random_state = 1)
```

```python
## se prprocesa las varibles x de entrenamiento y test
x_train1 = preprocessing.scale(x_train1)

x_test1 = preprocessing.scale(x_test1)
```

```python
## se carga 
x_train1.shape
```

```python
# creacion de  neural network
model = Sequential()
model.add(Dense(25, input_dim=Auto_fs.shape[1], activation='relu')) # capa 1
model.add(Dense(10, activation='relu')) # capa 2
model.add(Dense(1)) # salida
model.compile(loss='mean_squared_error', optimizer='adam')
monitor = EarlyStopping(monitor='val_loss', min_delta=1e-3, 
                        patience=5, verbose=1, mode='auto', 
                        restore_best_weights=True)
model.fit(x_train1,y_train1,validation_data=(x_test1,y_test1),
          callbacks=[monitor],verbose=2,epochs=1000)
```

```python
## se ve que cual es la mejor epoca y que la mejor epoca es la 60 se obitene el mejor calor 
```

```python
#Para visualizar la neural network
model.summary()
```

```python
history = model.fit(x_train1, y_train, epochs=10, batch_size=32,validation_split=0.15,validation_data=None,verbose=1)
```

```python
# evlacuion del Measure MSE error.
from sklearn import metrics
# Predict
pred = model.predict(x_test1)
score = metrics.mean_squared_error(pred,y_test1)
print("Final score (MSE): {}".format(score))
```

```python
#evaluacion de Root Mean Square Error
# Measure RMSE error.  RMSE is common for regression.
score = np.sqrt(metrics.mean_squared_error(pred,y_test1))
print("Final score (RMSE): {}".format(score))
```

```python
# Grafico de Regression .
def chart_regression(pred, y, sort=True):
    t = pd.DataFrame({'pred': pred, 'y': y.flatten()})
    if sort:
        t.sort_values(by=['y'], inplace=True)
    plt.plot(t['y'].tolist(), label='expected')
    plt.plot(t['pred'].tolist(), label='prediction')
    plt.ylabel('output')
    plt.legend()
    plt.show()
```

```python
# separacion de data en set de entrenamiento y testing, con parametros data y tipo de especie
X_train2,X_test2,y_train2,y_test2=train_test_split(Auto_fs,y_fs, random_state=0)
```

```python
# se importa funcion de libreria sklearn para KNN

from sklearn.neighbors import KNeighborsClassifier
```

```python
# creacion de variable knn para clasificador, k=5 (con k: numero de vecinos cercanos a considerar)
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
# entrenamiento de la variable knn, con parametros X_train2 e y_train2
knn.fit(X_train2,y_train2)
```

### Prediccion del Algoritmo

```python
# funcion score para ver que tan bien aprende el algoritmo ( en %), con parametros x_test2 e y_test2
knn.score(X_test2,y_test2)
```

### Pruebas de prediccion del algoritmo

```python
# algoritmo predice a que elemento pertenece con numeros aleatorios de entrada, donde:
# 0 = 1bbl
# 1 = 2bbl
# 2 = 4bbl
# 3 = idi
# 4 = mfi
# 5 = MPfi
# 6 = spdi
# 7 = spfi
knn.predict([[1,0,0,3,1,0,192,71,55,2954,3,2,131,120,15840]])
```

##### el algoritmo predice de acuerdo a los valores de entrada que la flor pertenece a la categoria 5 ( correspondiente a MPfi)

### Matriz de Confusion

```python
from sklearn.metrics import confusion_matrix
# se crea la variable pred para usarla al mostrar la matriz de confusion
pred = knn.predict(X_test2)
print('la matriz de confusion es:')

print(confusion_matrix(y_test2, pred))
```

### Grafica Matriz de confusion

```python
# se guarda la matriz de confusion en la variable mc, para luego graficarla con la funcion sns de la libreria seaborn
mc=confusion_matrix(y_test2, pred)
grafica=sns.heatmap(mc,annot=True)
grafica.set(xlabel='Verdaderos',ylabel='Predicciones')
```

### Metrica de desempe√±o: Precision ( confiabilidad del modelo)

```python
# se importa funcion de libreria sklearn para conocer el precision del algoritmo
from sklearn.metrics import accuracy_score

print('accuracy score', accuracy_score(y_test2, pred))
```

### Metrica de desempe√±o: Recall ( que tan bien detecta una clase o categoria)

```python
# se importa funcion de libreria sklearn para conocer el recall del algoritmo
from sklearn.metrics import classification_report

print(classification_report(y_test2, pred))
```

### Eleccion del mejor valor de k ( k optimo) 

```python
# Se crean las variables "k_ranqe" (para eje x) y "scores" (para eje y) las que seran usadas para graficar la exactitud del algoritmo. 
# Se crea un ciclo for para recorrer los valores que tomara k ( del 1 al 20 en este caso)

k_range = range(1, 20)
scores = []
for k in k_range:
    knn1 = KNeighborsClassifier(n_neighbors = k)
    knn1.fit(X_train2, y_train2)
    scores.append(knn1.score(X_test2, y_test2))
plt.figure()
plt.xlabel('k')
plt.ylabel('accuracy')
plt.scatter(k_range, scores)
plt.xticks([0,5,10,15,20])
```

##### Segun la informacion arrojada por la grafica se muestra que el mejor valor de k, al rango de valores [0,9] donde el algoritmo trabaja con un precision que va aumentando a medida de k aumenta estando cerca del 77%-78%-79% de exactitud, por lo que los valores de k optimo seran:  9,10,11,12,13,14,15......
