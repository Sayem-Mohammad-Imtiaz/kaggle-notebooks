```python
df = pd.read_csv("../input/star-type-classification/Stars.csv")
df.head()
```

```python
df.shape
```

```python
data = df.sample(frac=0.9, random_state=786).reset_index(drop=True)
data_unseen = df.drop(data.index).reset_index(drop=True)

print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.classification import *
```

```python
clf = setup(data = data, target = 'Type', session_id=123) 
```

```python
compare_models()
```

```python
dt = create_model('dt')
```

```python
best_model = automl()
```

```python
best_model
```

```python
print(dt)
```

```python
tuned_dt = tune_model(dt)
```

```python
plot_model(tuned_dt, plot = 'confusion_matrix')
```

```python
plot_model(tuned_dt, plot = 'class_report')
```

```python
evaluate_model(tuned_dt)
```

```python
predict_model(tuned_dt)
```

```python
final_dt = finalize_model(tuned_dt)
```

```python
print(final_dt)
```

```python
unseen_predictions = predict_model(final_dt, data=data_unseen)
unseen_predictions.head(10)
```
