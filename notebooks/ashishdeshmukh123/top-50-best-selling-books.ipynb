***We will Review Overall Top 50 books***

```python
df=books[['Name','Reviews']].groupby('Name',axis=0).sum().sort_values('Reviews',ascending=False)[0:50]
a=df
a=a.sort_values('Reviews')
```

```python
plt.figure(figsize=(40,60))
b=plt.barh(a.index,a.Reviews,height=0.5,color='m',edgecolor='black',align='center')
plt.bar_label(b,label=a.Reviews,padding=8, color='b', fontsize=30)
plt.title("Top 50 Best Selling Books \n")
plt.xlabel("Sales",fontsize=30)
plt.text(0.1,-1,"From Above Plot We Can Observe Top 50 Best Seller Books of all Time",fontsize=30)
plt.legend()
plt.tight_layout()
plt.rcParams.update({'font.size': 30})
```

 **Top 50 Best Selling Books Every Year ***

```python
def Plot(a,year):
    a=a[a['Year']==year]
    df=a[['Name','Reviews']].groupby('Name',axis=0).sum().sort_values('Reviews',ascending=True)[0:50]
    a=df
    a=a.sort_values('Reviews',ascending=False)
    
    plt.figure(figsize=(50,60))
    b=plt.barh(df.index,df.Reviews,height=0.5,color='m',edgecolor='black',align='center')
    plt.bar_label(b,label=a.Reviews,padding=2, color='b', fontsize=40)
    plt.rcParams.update({'font.size': 40})
    plt.title(f"\n \n Top 50 Best selling books in {year} \n")
    
    plt.legend()
    plt.tight_layout()
    
    plt.show()    
```

```python
k=[2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019]
for i in k:
    Plot(books,i)
```
