### Start Time

```python
df.Start_Time
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time[0]
```

```python
sea.distplot(df.Start_Time.dt.hour, bins=24, kde=False, norm_hist=True)
```

* A high percentage of accidents happen between 6 am to 9 am (probably working hours).
* Next highest percentage is between 3pm to 6pm.

```python
sea.distplot(df.Start_Time.dt.dayofweek, bins=7, kde=False, norm_hist=True)
```

```python
sunday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sea.distplot(sunday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

On Sundays, the peek occur in 10 am to 3pm, unlike weekdays.

```python
monday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sea.distplot(monday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

```python
saturday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek == 5]
sea.distplot(saturday_start_time.dt.hour, bins=24, kde=False, norm_hist=True)
```

Same with Saturdays, The peak time of accidents happening is between 10 pm to 4pm. 

```python
df_2019 = df[df.Start_Time.dt.year == 2019]
sea.distplot(df_2019.Start_Time.dt.month, bins=12, kde=False, norm_hist=True)
```

More data is missing for 2016. Maybe even for 2017
