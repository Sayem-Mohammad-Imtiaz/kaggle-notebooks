```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv("../input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv")
df
```

```python
df.head()
```

```python
df.drop("id", axis = 1, inplace = True)
```

```python
df.dropna(inplace = True)
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True,annot_kws={"size":15})
```

```python
gender = pd.get_dummies(df["gender"], drop_first = True)
ever_married = pd.get_dummies(df["ever_married"], drop_first = True)
work_type = pd.get_dummies(df["work_type"], drop_first = True)
Residence_type = pd.get_dummies(df["Residence_type"], drop_first = True)
smoking_status = pd.get_dummies(df["smoking_status"], drop_first = True)
```

```python
df.drop(["gender", "ever_married", "work_type", "Residence_type", "smoking_status"], axis = 1, inplace = True)
```

```python
df = pd.concat([df, gender, ever_married, work_type, Residence_type, smoking_status], axis = 1)
```
