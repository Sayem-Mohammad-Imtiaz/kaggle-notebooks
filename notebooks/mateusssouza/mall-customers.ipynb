```python
# copy the dataframe:
customers_with_null = customers
# customers.loc[1,'Age'] - returns the 'Age' attribute of the first instance
# customers.sample(frac=0.1) - randomly returns 10 percent of instances
# np.nan = returns nan (not a number)
for col in customers_with_null.columns:
    customers_with_null.loc[customers_with_null.sample(frac=0.1).index,col] = np.nan   
```

```python
objects_in_customers = customers.select_dtypes(include=['object']);
objects_in_customers.head(10)
```

```python
objects_in_customers.info();
```

**Aplicando o mapeamento**

```python
replace_map = {'Gender':{'Male':1,'Female':2}} # defines the dictionary to be used
labels = objects_in_customers['Gender'].astype('category').cat.categories.tolist();
replace_map_comp = {'Gender':{k: v for k,v in zip(labels,list(range(1,len(labels)+1))) }}

print(replace_map_comp)
```

```python
customers_replace = pd.read_csv('../input/customer-segmentation-tutorial-in-python/Mall_Customers.csv')
```

```python
customers_replace.replace(replace_map_comp,inplace=True)
customers_replace.head()
```

```python
customers_replace.info();
```
