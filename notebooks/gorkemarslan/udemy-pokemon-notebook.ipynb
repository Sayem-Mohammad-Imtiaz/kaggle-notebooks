```python
def return_a_power_function_of_power(power=2):
    """default power value is 2"""
    def retfunc(*numbers):
        if len(numbers)>1:
            return tuple(map(lambda x:x**power,numbers))
        return numbers[0]**power
    return retfunc
cube=return_a_power_function_of_power(3)
cube(10),cube(5,6,7)
```

```python
gender=["F","M","M","F"]
weight=[57,85,69,71]
weighertype=[1,2,1,2]
colnames=["gender","weight","weightertype"]
valslist=[gender,weight,weighertype]
zipped=zip(colnames,valslist)
dicted=dict(zipped)
df=pd.DataFrame(dicted)
df.set_index(["gender","weightertype"],inplace=True)
print(df,end="\n--------------------------\n")
df.sort_index(inplace=True)
print(df,end="\n--------------------------\n")
df=df.unstack("gender")
print(df,end="\n--------------------------\n")
df=df.stack("gender")
print(df,end="\n--------------------------\n")
print(df.index.name,df.index.names,end="\n--------------------------\n")
df=df.swaplevel(0,1)
print(df,end="\n--------------------------\n")
```
