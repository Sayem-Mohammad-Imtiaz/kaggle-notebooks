## EDA: Разведочный анализ данных (Exploratory Data Analysis)

```python
df = pd.read_csv('/kaggle/input/videogamesales/vgsales.csv')
df.head()
```

```python
print(df.info())

#заполним пропущенные данные. 
#также их можно удалить при помощи dropna()
df.fillna(0, inplace=True) 

print(df.info())
```

```python
# какие платформы самые популярные?
df_sales = df[['Platform', 'Global_Sales']].groupby('Platform').agg('count').nlargest(15, 'Global_Sales').reset_index()


df_sales.plot.bar(x = 'Platform', y = 'Global_Sales', rot = 90, figsize = (8, 4), alpha = 0.6, legend = False)
plt.xlabel('Платформа')
plt.ylabel('Продажи в мире')
plt.title("Зависимость продаж от выбора платформы");
```

```python
# какие жанры в мире самые популярные?
df_genre = df[['Genre', 'Global_Sales']].groupby('Genre').agg('count').nlargest(15, 'Global_Sales').reset_index()


df_genre.plot.bar(x = 'Genre', y = 'Global_Sales', rot = 90, figsize = (8, 4), alpha = 0.6, legend = False)
plt.xlabel('Жанр')
plt.ylabel('Продажи в мире')
plt.title("Зависимость жанра от выбора платформы");
```

```python
cols = ['NA_Sales', 'EU_Sales', 'Year', 'Other_Sales', 'JP_Sales', 'Global_Sales']

heatmap = sns.heatmap(df[cols].corr(),
                 cbar=True,
                 annot=True)
```

# Регрессионный анализ

### Линейная регрессия

```python
# Разделяем переменные
X = df[['NA_Sales']].values
y = df['Global_Sales'].values

# Разбиваем данные на тестовые и тренировочные
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=0)


# Модель
slr = LinearRegression()  

# Обучаем модель на тренировочных данных
slr.fit(X_train, y_train) 

# Строим прогноз на тренировочных и тестовых данных
y_train_pred = slr.predict(X_train)
y_test_pred = slr.predict(X_test)           

# Отображаем результаты
plt.scatter(X_train, y_train)
plt.title('Результаты на тренировочных данных')
plt.xlabel('NA_SALES'),plt.ylabel('GLOBAL_SALES') 
plt.plot(X_train, y_train_pred, color='red', linewidth=2);
```

```python
# Оценка
train_mse =  mean_squared_error(y_train, y_train_pred)
test_mse = mean_squared_error(y_test, y_test_pred)

print(f'MSE train: {round(train_mse, 3)}, test: {round(test_mse, 3)}')
```

```python
# Отображаем результаты для тестовых данных
plt.scatter(X_test, y_test)
plt.title('Результаты на тестовых данных')
plt.xlabel('NA_SALES'),plt.ylabel('GLOBAL_SALES') 
plt.plot(X_test, y_test_pred, color='red', linewidth=2);
```
