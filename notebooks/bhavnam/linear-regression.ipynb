# **Linear Regression with a single variable**
Simple linear regression is a statistical approach for modelling the relationship between a predictor variable X and a response variable Y. It assumes there is a linear relationship between these two variables and we use that to predict a quantitative output.

```python
sns.barplot(x=houses['sqft_living'], y=houses['price'])
#plt.xticks(rotation= 45)
plt.xlabel('Area')
plt.ylabel('Price')
```

```python
plt.figure(figsize=(10,10))
sns.jointplot(x=houses['sqft_living'], y=houses['price'], data=houses, kind='reg')
```

```python
sns.jointplot(houses.sqft_living, houses.price, kind="kde")
```

```python
cols = ['sqft_living']
x = houses[cols]
y = houses.price.values
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2) 
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
print (lr.intercept_)
print (lr.coef_)
```

# **Visualising the output** 

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_test, lr.predict(x_test), color = 'blue')
plt.xlabel('Area (in sqft)')
plt.ylabel('Price')
plt.show()
```

```python
from sklearn.metrics import r2_score

r = r2_score(y_test, lr.predict(x_test))
mse = mean_squared_error(y_test, lr.predict(x_test))

print(r)
print(np.sqrt(mse))
print(cross_val_score(lr, x , y, cv=3))
```

# Ordinary Least Squares Regression

```python
import statsmodels.api as sm

X = sm.add_constant(x)  # Adds a constant term to the predictor
X.head()

est=sm.OLS(y, X)
est = est.fit()
est.summary()
```

# Multivariable Linear Regression    
**Modelling**  
Just like for simple linear regression, we will define our features and target variable and use scikit-learn library to perform linear regression.

```python
cols = ['grade','sqft_living','floors','bedrooms']
x=np.array(houses[cols])
y=np.array(houses['price'])
```

```python
plt.figure(figsize=(10,10))
sns.jointplot(x=houses['grade'], y=houses['price'], data=houses, kind='reg')
```

```python
plt.figure(figsize=(10,10))
sns.jointplot(x=houses['floors'], y=houses['price'], data=houses, kind='reg')
```

```python
plt.figure(figsize=(10,10))
sns.jointplot(x=houses['bedrooms'], y=houses['price'], data=houses, kind='reg')
```

```python
lr=LinearRegression()
lr=lr.fit(x,y)
y_pred=lr.predict(x)

rmse=np.sqrt(mean_squared_error(y,y_pred))
r2=lr.score(x,y)
print(rmse)
print(r2)
```
