# Categorical Variable Analysis

```python
def bar_plot(variable):
    '''
    input: variable ex: genre, author
    output: bar plot and value count
    '''
    var = df[variable]
    varCount = var.value_counts()
    
    plt.figure(figsize=(9,3))
    plt.bar(varCount.index, varCount)
    plt.xticks(varCount.index, varCount.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
    print(f"{variable} \n{varCount}")
```

```python
category1 = ["Genre"]
for c in category1:
    bar_plot(c)
```

```python
authors = df["Author"]
authors = authors[0:25]
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)
plt.ylabel("num of publish")
plt.xlabel("authors")
plt.title("Top 25 published authors")
sns.barplot(x=authors.index, y=authors.values)
```

# Numerical Variable Analysis

```python
def plot_hist(variable, df):
    plt.figure(figsize=(18,6))
    plt.hist(df[variable], bins=50)
    plt.xlabel(variable)
    plt.ylabel("Frequency")
    plt.title(f"{variable} distribution")
```

```python
numericVars = ["Price", "User Rating", "Reviews"]
for var in numericVars:
    plot_hist(var, df)
```
