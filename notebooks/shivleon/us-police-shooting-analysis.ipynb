# Basic Information

```python
def basic_info(data):
    print(data.shape)
    print(data.size)
    print(data.info())
    cat, num = list(), list()
    for i in data.columns:
        if data[i].dtype == object:
            cat.append(i)
        else:
            num.append(i)
    print(cat, "\n")
    print(num)
    return cat, num
```
