```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('../input/imdb-5000-movie-dataset/movie_metadata.csv')
movies.head(3)
```

## Best Non-English Low-Budget Movies Barchart

```python
non_english_movies = movies.loc[(movies.language != 'English') & (movies.budget < 10000000.0)]
country_scores = non_english_movies.groupby('country').imdb_score.mean()
country_scores.sort_values(ascending=False).plot.bar()
type(country_scores)
```

## Movies Length vs Rating

```python
movies.loc[:, ['duration', 'imdb_score']].plot.scatter(x='duration', y='imdb_score')
```
