```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from scipy.stats import zscore
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df = pd.read_csv('/kaggle/input/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
df.shape
```

# Dropping Cutomer ID Column

```python
df.drop('customerID',axis=1,inplace=True)
```

# Summary of data

```python
df.describe()
```

```python
df.info()
```

# Checking & Dropping Duplicates

```python
dups=df.duplicated()
df[dups]
```

```python
df.drop_duplicates(inplace=True)
```

# Dropping Null Values

```python
df['TotalCharges']=pd.to_numeric(df["TotalCharges"],errors="coerce")
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

# Categorical Column

```python
for column in df.columns:
    if df[column].dtype == 'object':
        print(column.upper(),': ',df[column].unique())
        print(df[column].value_counts().sort_values())
        print('\n')
```

# ENCODING CATEGORICAL DATA

```python
for feature in df.columns: 
    if df[feature].dtype == 'object': 
        print('\n')
        print('feature:',feature)
        print(pd.Categorical(df[feature].unique()))
        print(pd.Categorical(df[feature].unique()).codes)
        df[feature] = pd.Categorical(df[feature]).codes
```

```python
column=['gender', 'Partner', 'Dependents','PhoneService', 'PaperlessBilling']
```

```python
df =pd.get_dummies(df, columns=column,drop_first=True)
df.head()
```

```python
df.shape
```

```python
df.info()
```

#  Uni-Variate Analysis

# BOXPLOTS

```python
continuous=df.dtypes[(df.dtypes=='int64')].index
df_plot=df[continuous]
fig=plt.figure(figsize=(20,20))
for i in range(0,len(df_plot.columns)):
    ax=fig.add_subplot(4,3,i+1)
    sns.boxplot(y=df_plot[df_plot.columns[i]])
    ax.set_title(df_plot.columns[i],color='red')
    plt.grid()
plt.tight_layout()
```

```python
continuous=df.dtypes[(df.dtypes=='float64')].index
df_plot=df[continuous]
fig=plt.figure(figsize=(20,20))
for i in range(0,len(df_plot.columns)):
    ax=fig.add_subplot(4,3,i+1)
    sns.boxplot(y=df_plot[df_plot.columns[i]])
    ax.set_title(df_plot.columns[i],color='red')
    plt.grid()
plt.tight_layout()
```

# DIST PLOTS

```python
continuous=df.dtypes[(df.dtypes=='int64')].index
df_plot=df[continuous]
fig=plt.figure(figsize=(20,20))
for i in range(0,len(df_plot.columns)):
    ax=fig.add_subplot(4,3,i+1)
    sns.distplot(df_plot[df_plot.columns[i]],kde=True)
    ax.set_title(df_plot.columns[i],color='red')
    plt.grid()
plt.tight_layout()
```
