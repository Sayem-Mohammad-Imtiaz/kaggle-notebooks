#### 5- Data Preprocessing

```python
dataset.drop(columns=['User ID'], axis=1, inplace=True)
```

```python
# Converting Gender into Dummy Data
gender = pd.get_dummies(dataset['Gender'], drop_first=True)


# Adding dummy data to data set and removing Gender column
datset = pd.concat([gender, dataset], axis=1)
dataset.drop(columns=['Gender'], axis=1, inplace=True)
```

#### 6-  Extracting Feature

```python
X = dataset.iloc[:, :-1].values    #Features
y = dataset.iloc[:, -1].values     #Targets
```

#### 7- Feature Scaling

```python
# from sklearn.preprocessing import StandardScaler
```

```python
st = StandardScaler()
```

```python
X = st.fit_transform(X)
```

```python
X
```

#### 8- Splitting data into training and testing

```python
# from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# from sklearn.svm import SVC
```

```python
# Using Polynomial Kernel

poly_model = SVC(kernel='poly', degree=3)
```

```python
poly_model.fit(X_train, y_train)
```

```python
poly_model.score(X_train, y_train)
```

```python
poly_model.score(X_test, y_test)
```

```python
poly_model.predict(st.transform([[19,19000]]))
```

```python
y_pred_poly = poly_model.predict(X_test)
```

```python
# from sklearn.metrics import confusion_matrix, classification_report
```

```python
cf_matrix = confusion_matrix(y_test, y_pred_poly)
sns.heatmap(cf_matrix, annot=True)
plt.show()
```

```python
print(classification_report(y_test, y_pred_poly))
```

```python
# Using RBF Kernel
```

```python
rbf_model = SVC(kernel='rbf').fit(X_train, y_train)
```

```python
# Score on test data

rbf_model.score(X_test, y_test)
```

```python
# Score on training data

rbf_model.score(X_train, y_train)
```

```python
rbf_model.predict(st.transform([[26,43000]]))
```

```python
y_pred_rbf = rbf_model.predict(X_test)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred_rbf), annot=True)
plt.show()
```

```python
print(classification_report(y_test, y_pred_rbf))
```

### Random Forest Classifier
