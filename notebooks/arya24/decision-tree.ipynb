```python
from sklearn.preprocessing import LabelEncoder

df1 = df.copy(deep = True)
le_color = LabelEncoder()
df1['Favorite Color'] = le_color.fit_transform(df['Favorite Color'])

le_genre = LabelEncoder()
df1['Favorite Music Genre'] = le_genre.fit_transform(df['Favorite Music Genre'])


le_beverage = LabelEncoder()
df1['Favorite Beverage'] = le_beverage.fit_transform(df['Favorite Beverage'])


le_drink = LabelEncoder()
df1['Favorite Soft Drink'] = le_drink.fit_transform(df['Favorite Soft Drink'])


le_gender = LabelEncoder()
df1['Gender'] = le_gender.fit_transform(df['Gender'])
```

```python
#Dependent variables
X = df1.iloc[:, 0:4]

#Independent variables
y = df1.iloc[:, 4]
```
