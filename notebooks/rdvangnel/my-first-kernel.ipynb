```python
#correlation map
f,ax = plt.subplots(figsize=(10,10))
sns.heatmap(df.corr(),annot =True,linewidths=5,fmt= '.1f',ax=ax)
plt.show()
```

> First of all, I prepared the Correlation Map.
> >  Then I checked therelationships between 'Open' with 'Close' .(Correlation Rate=1)
> > > And I chose to analyze the process between 10000 and 5000.


```python
x=df[(df['Open']>5000)&(df['Open']<10000)&(df['Close']<10000 )&(df['Close']>5000)]
```

```python
x.head(10)
```

* No fluctuation in the range of 3,6 to 3,8.

```python
# Line Plot
# color = color, label = label, linewidth = width of line, alpha = opacity, grid = grid, linestyle = sytle of line
x.Open.plot(kind = 'line', color = 'g',label = 'Open',linewidth=5,alpha = 0.5,grid = True,linestyle = ':')
x.Close.plot(color = 'r',label = 'Close',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('x axis')              # label = name of label
plt.ylabel('y axis')
plt.title('Line Plot')            # title = title of plot
plt.show()
```

* 'Open' and 'Close' are in direct proportion

```python
# Scatter Plot 
# x = Open, y = Close
x.plot(kind='scatter', x='Open', y='Close',alpha = 0.9,color = 'red')
plt.xlabel('Open')              # label = name of label
plt.ylabel('Close')
plt.title('Open Close Scatter Plot')            # title = title of plot
plt.show()
```

The Highest Timestamp has been observed between 6000 and 7000.

```python
# Histogram
# bins = number of bar in figure
x.Open.plot(kind = 'hist',bins = 120,figsize = (12,12),color="r")
plt.show()
```
