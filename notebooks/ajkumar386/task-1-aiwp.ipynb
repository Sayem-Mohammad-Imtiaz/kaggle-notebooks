# TASK 1
Train a linear regression classifier to fit the data. Use Salary as a target column (y). Form the models with following configurations
* (a) 50% Train, 50% Test
* (b) 70% Train, 30% Test
* (c) 80% Train, 20% Test

Analyze the MSE, RMSE for each configuration. Write your own function to determine MSE/RMSE.

Crosscheck it with the results given by sklearn metrics.

# Input Dataset

```python
import numpy as np # for linear algebra
import pandas as pd # data processing, CSV file I/O
data = pd.read_csv('../input/salary-data-simple-linear-regression/Salary_Data.csv')
data.head()
```

# Function for MSE

```python
def mse(predictions, targets):
    return  (((predictions - targets) ** 2).mean())
```

# Function for RMSE

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())
```

# TASK A
> For 50% Train and 50% Test

```python
x = data.iloc[:, :-1] # YearsExperience as x
y = data.iloc[:, [-1]] # Salary as y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size = 0.50,test_size = 0.50) # 50% Train and 50% Test
```

```python
from sklearn import linear_model
lr = linear_model.LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
```

> # Using MSE and RMSE functions

```python
print('Mean squared error using fuction: %.2f' % mse(y_test,y_pred))
print('Root mean squared error using fuction: %.2f' % rmse(y_test, y_pred))
```

>  # Cross-checking MSE & RMSE using sklearn metrics

```python
from sklearn.metrics import mean_squared_error
print('Mean squared error using sklearn: %.2f' % mean_squared_error(y_test,y_pred))
rms = np.sqrt(mean_squared_error(y_test, y_pred))
print('Root mean squared error using sklearn: %.2f' % rms)
```

***Hence, Verified***
