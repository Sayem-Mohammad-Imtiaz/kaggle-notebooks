Create 4 new dataframes with each regions gross.

```python
top_US = top_games[[ 'Name','Platform', 'Year', 'Genre', 'Publisher', 'NA_Sales']].sort_values(by = 'NA_Sales', ascending=False)
top_EU = top_games[[ 'Name','Platform', 'Year', 'Genre', 'Publisher', 'EU_Sales']].sort_values(by = 'EU_Sales', ascending=False)
top_JP = top_games[[ 'Name','Platform', 'Year', 'Genre', 'Publisher', 'JP_Sales']].sort_values(by = 'JP_Sales', ascending=False)
top_other = top_games[[ 'Name','Platform', 'Year', 'Genre', 'Publisher', 'Other_Sales']].sort_values(by = 'Other_Sales', ascending=False)
```

Lets go back and see what the most popular Publisher was.

```python
tp = top_games.groupby('Publisher')['Global_Sales'].sum().to_frame(name = "Sales")
tp.sort_values(by = 'Sales', ascending = False)[0:20].plot.bar()
```

Now lets look at the mst popular genres from North America and the EU. We can use similar code to what we used for finding the biggest publishers.

Now that we have the data we can make those into pandas datasets and compare.

```python
top_Genres = top_US.groupby('Genre')['NA_Sales'].sum().to_frame(name = "NA").sort_values(by = 'NA', ascending = False)
top_EU_Genres = top_EU.groupby('Genre')['EU_Sales'].sum().to_frame(name = "EU").sort_values(by = 'EU', ascending = False)
top_JP_Genres = top_JP.groupby('Genre')['JP_Sales'].sum().to_frame(name = "JP").sort_values(by = 'JP', ascending = False)
top_other_Genres = top_other.groupby('Genre')['Other_Sales'].sum().to_frame(name = "Other").sort_values(by = 'Other', ascending = False)

display(top_Genres)
```

Now lets add the data together.

```python
top_Genres['EU'] = top_EU_Genres
top_Genres['JP'] = top_JP_Genres
top_Genres['Other'] = top_other_Genres

top_Genres
```

Now we can look at them as graphs

```python
top_Genres.plot.bar()
```

Now lets try to scale the columns to make a more meaningful graph.  
Scaling data is important because it allows use to make certain conclusions. The unscaled data above is useful when considering the total output by area and how that affeects the total numbers, but it doesn't give us any meaningful conclusions about camparing regions. In fact, some one might come to a false conclusion when looking at the data above. That is why it is important to be aware of how a data set gives information and make sure that the data is being received properly.

```python
scale_NA = top_Genres['NA'].sum()
scale_EU = top_Genres['EU'].sum()
scale_JP = top_Genres['JP'].sum()
scale_other = top_Genres['Other'].sum()
stg = top_Genres["NA"].div(scale_NA)
scaledTG = stg.to_frame(name = "NA")
scaledTG['EU'] = top_Genres['EU'].div(scale_EU)
scaledTG['JP'] = top_Genres['JP'].div(scale_JP)
scaledTG['Other'] = top_Genres['Other'].div(scale_other)
scaledTG
```

Now we have a scaled dataframe, so we can better understand what certain regions like.

```python
scaledTG.transpose().plot.bar(stacked=True, figsize = (10,10))
```
