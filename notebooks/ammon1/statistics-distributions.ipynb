##Predictions

```python
df1=pd.get_dummies(df)
```

```python
fig = plt.subplots(figsize=(25, 25))
sns.heatmap(df1.corr(), annot=True, fmt=".1f")
```

```python
df1.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df_scaled = pd.DataFrame(scaler.fit_transform(df1), columns=df1.columns)
```

```python
X=df_scaled.iloc[:,3:].values
Y_m=df_scaled.iloc[:,0].values
Y_r=df_scaled.iloc[:,1].values
Y_w=df_scaled.iloc[:,2].values
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV

rfc=RandomForestRegressor(random_state=42)
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['mse','mae']
}
CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid, cv= 5)
CV_rfc.fit(X,Y_r)
CV_rfc.best_params_
```

```python
CV_rfc.score(X,Y_r)
```

```python
from sklearn.linear_model import Lasso


lasso=Lasso()
param_grid = { 
    'alpha': [0.01,0,1, 1,10],
    'max_iter':[100,1000,10000]
}
CV_lasso = GridSearchCV(estimator=lasso, param_grid=param_grid, cv= 5)
CV_lasso.fit(X,Y_r)
CV_lasso.score(X,Y_r)
```

Very low score of RF and Lasso sugest there are no rules of scoring math exam base on gender, race and other given parameters
