```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st
from scipy import stats

df = pd.read_csv('../input/avocado-prices/avocado.csv')

def descriptive_statistics(col):
    print(col)
    arr = df[col].to_numpy()
    print("- mean   : %.2f" % st.mean(arr))
    print("- median : %.2f" % st.median(arr))
    print("- modus  : %.2f" % stats.mode(arr)[0])
    print("- varian : %.2f" % st.variance(arr))
    print("- st_dev : %.2f" % st.stdev(arr))

def correlation_analysis(x,y,sh,cl,id):
    arrx = df[x].to_numpy()
    arry = df[y].to_numpy()
    descriptive_statistics(x)
    descriptive_statistics(y)
    print("Pearson correlation coefficient : %.2f" % np.corrcoef(arrx, arry)[0][1])
    plt.plot(arrx, arry, sh, color=cl);
    plt.show()

correlation_analysis('Total Bags','Small Bags','+','red',1)
correlation_analysis('Total Volume','Large Bags','o','green',2)
correlation_analysis('AveragePrice','XLarge Bags','^','blue',3)
```
