> **NESTED FUNCTİON**

```python
def func():
    """return value x*add"""
    def add():
        """add local variable """
        x=2
        y=8
        z=y+x
        return z
    return add()**2
print(func())
```

> **DEFAULT and FLEXIBLE ARGUMENTS**

```python
#default arguments

def func(x,y,z=3):      # default argument is overwritten
    """ return x+(y*z) """
    return z+(x*y)
print(func(2,1))
```

```python
#flexible arguments

def func2(*args):
    for i in args:
        print(i)
func2(1,2,3,4,5)

#flexible arguments **kwargs --->> dictionary
def func3(**kwargs):
    for key, value in kwargs.items():
        print(key+":"+value)
        
func3(alex="Brazil",hagi="Romania")

```
