## Problem Statement

### Predicting students will get placement or not

#### Feature Selection: <br> from above Data Analysis we are select independent or feature variables are given below
- gender
- ssc_p (SSC Percentage)
- hsc_p (HSC Percentage)
- hsc_s (HSC Specialisation)
- degree_p (Degree Percentage)
- degree_t (Degree Type)
- workex (Work Experience)
- specialisation

##### target variable is status

```python
# gathering for status prediction
status_data = main_data[["gender", "ssc_p", "hsc_p", "hsc_s", "degree_p", "degree_t", "workex", "specialisation", "status"]].copy()
status_data.sample(5)
```

#### Encode Text Data into numeric format

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
status_data["gender"] = le.fit_transform(status_data.gender) # 0 = female, 1 = male
status_data["hsc_s"] = le.fit_transform(status_data.hsc_s) # 0 = Arts, 1 = Commerce, 2 = Science
status_data["degree_t"] = le.fit_transform(status_data.degree_t) # 0 = Comm&Mgmt, 1 = Others, 2 = Sci&Tech
status_data["workex"] = le.fit_transform(status_data.workex) # 0 = No, 1 = Yes
status_data["specialisation"] = le.fit_transform(status_data.specialisation) # 0 = Mkt&Fin, 1 - Mkt&HR
status_data["status"] = le.fit_transform(status_data.status) # 0 = Not Placed, 1 = Placed 
```

```python
status_data.head(2)
```

##### creating feature Matrix X and resonse vector y

```python
X = status_data.loc[:, status_data.columns != "status"]
y = status_data.status
```

```python
# split data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 15)
```

#### Predicting status is a classification problem so here we use Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
logreg.score(X_test, y_test)
```

### verfying prediction using confusion matrix

```python
y_pred = logreg.predict(X_test)
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(y_test, y_pred)
sns.heatmap(cf, annot = True)
plt.xlabel("Predicted")
plt.ylabel("Actual")
```

#### we successfully created our model with 90% accuracy <br> from above heat map we get below information
- In 1 case student was placed but our model said he doesn't placed
- In 3 case students weren't placed but our model said  they does placed
