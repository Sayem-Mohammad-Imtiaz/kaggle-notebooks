Assume the minimum wage in Turkey 2000 per month then we can calculate the monthly deficit for each capita so that it can be observed that wether there is no effect of money on suicides or not.

```python
result = map(lambda x : int((16000-x)/12),data[data["country"] == "Turkey"]["gdp_per_capita ($)"])
print(list(result))
```

Let's conjugate country-year and generation.

```python
z = zip(data["country-year"],data["generation"])
data["new"] = list(z)
print(data.new)
```

Now we will look at the ratios that bigger than 2.5 using list comprehension method.

```python
data["statementofCountries"] =["problematic" if i >2.5 else "normal" if i==2.5 else "better" for i in data["suicides/100k pop"]]
print(data[data["statementofCountries"]=="problematic"]["country-year"])
```
