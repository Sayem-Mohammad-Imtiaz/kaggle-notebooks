```python
import numpy as np
import glob,os
import scipy.io.wavfile
from matplotlib import pyplot as plt
from IPython.display import HTML, display
import time
```

```python
def progress(value,master, max=100):
    return HTML(str(value)+" - "+master+"""
        <progress
            value='{value}'
            max='{max}',
            style='width: 100%'
        >
            {value}
        </progress>
    """.format(value=value, max=max))
```

```python
mono_count = 0
stereo_count = 0

count=0
out = display(progress(0, "master",21000), display_id=True)
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        file = os.path.join(dirname, filename)
        
        if(file.find(".wav")>0):
            try:
                out.update(progress(count,file, 21000))
                rate,data = scipy.io.wavfile.read(file)
                dim=data.ndim
                if(dim==1):
                    scipy.io.wavfile.write("/kaggle/working/Audio/UniqueAudio-Mono-"+str(rate)+"-"+str(count)+".wav",rate,data)
                else:
                    scipy.io.wavfile.write("/kaggle/working/Audio/UniqueAudio-Stereo-"+str(rate)+"-"+str(count)+".wav",rate,data)
                count = count + 1
            except:
                print(count,end=", ")
```
