```python
sp_data = pd.read_csv('../input/student-performance-data-set/student-por.csv')
sp_data.head()
```

```python
sp_data.columns
```

```python
sp_data = sp_data.drop(columns = ['sex','school','internet', 'famsize','address', 'Pstatus', 'Medu', 'Fedu', 'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime','schoolsup', 'famsup', 'paid', 'activities', 'nursery', 'higher','G1', 'G2','Dalc', 'Walc', 'health', 'romantic', 'famrel'])

sp_data.head()
sp_data.shape
```

X will be all colums and y will be G3 grades on a scale of 0-20

```python
#import some important libraries
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split


#ML using one decision tree
#Create target object y
y = sp_data.G3

#Create X
features = [ 'age', 'studytime', 'failures', 'freetime', 'goout', 'absences']
X = sp_data[features]

#split model
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)

#defining model
sp_data_model = DecisionTreeRegressor(random_state=1)

#fit model
sp_data_model.fit(train_X, train_y)


#predict
val_predict = sp_data_model.predict(val_X)

#calculate MAE
val_mae = mean_absolute_error(val_predict, val_y)
print("Val MAE for DecisionTreeRegressor: " )
print(val_mae)
```

```python
#Finding best value for leaf nodes
#paste in get_mae funtion
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)

#now use it and a loop to try and find best leaf nodes to use
for max_leaf_nodes in [2,3,4,5,50,500,5000,50000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```

```python
#Now to use a Random Forrest and compare MAE's
#define model
rf_sp_data_model = RandomForestRegressor(random_state = 1)

#fit
rf_sp_data_model.fit(train_X, train_y)

#predict
pred = rf_sp_data_model.predict(val_X)

#calculate MAE
rf_val_mae = mean_absolute_error(val_y, pred)

print("Validation MAE for RandomForestRegressor: ")
print(rf_val_mae)
```
