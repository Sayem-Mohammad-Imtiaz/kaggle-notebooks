# Generating wordcloude image

```python
# importing required librarys
from nltk.corpus import stopwords
from spacy.lang.en.stop_words import STOP_WORDS
from wordcloud import WordCloud
# Create stopword list:
STOP_WORDS.add('otter')
stopwords = set(list(STOP_WORDS) +list(stopwords.words()))
stopwords.update(["br", "href", 'https'])
stopwords.update(stopwords)
textt = " ".join(desc for desc in zomatoData["dish_liked"])
wordcloud = WordCloud(stopwords=stopwords).generate(textt)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title('Most Liked Dishes',fontsize=20,fontweight='bold')
plt.savefig('wordcloud11.png')
plt.show()
```

```python
# Droping unnessesory columns from datasets
Zomato=zomatoData.drop(['url','dish_liked','phone'] ,axis=1)
```

```python
# Droping duplicate data from datasets
Zomato.duplicated().sum()
Zomato.drop_duplicates(inplace=True)
```

```python
# printing columns name
Zomato.columns
```

# Data cleaning

```python
# renaming  column names
Zomato=Zomato.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'type','listed_in(city)':'city'})
Zomato.columns
```

```python
# relpacing cost string to floats
Zomato['cost']=Zomato['cost'].astype(str)
Zomato['cost']=Zomato['cost'].apply(lambda x:x.replace(",","."))
Zomato['cost']=Zomato['cost'].astype(float)
Zomato.info()
```

```python
#  printing unique rating
Zomato['rate'].unique()
```

```python
# Removing '/5' from  dataset
Zomato=Zomato.loc[Zomato.rate!='NEW']
Zomato=Zomato.loc[Zomato.rate!='-'].reset_index(drop=True)

Zomato['rate']=Zomato['rate'].astype(str)
Zomato['rate']=Zomato['rate'].apply(lambda x:x.replace("/5",''))
Zomato['rate']=Zomato['rate'].astype(float)
Zomato['rate'].unique()
```

```python
# standardlizing 
Zomato.name=Zomato.name.apply(lambda x:x.title())
Zomato.online_order.replace(("Yes","No"),("True","False"),inplace=True)
Zomato.book_table.replace(("Yes","No"),("True","False"),inplace=True)
Zomato.cost.unique()
```

```python
Zomato
```

```python
# encoding  data
def Encoder(Zomato):
    for column in Zomato.columns[~Zomato.columns.isin(['rate','cost','votes'])]:
        Zomato[column]=Zomato[column].factorize()[0]
    return Zomato
Zomato_encoded=Encoder(Zomato.copy())
```

```python
Zomato_encoded
```

# correlations

```python
corr=Zomato_encoded.corr(method='kendall')
plt.figure(figsize=(15,8))
sn.heatmap(corr,annot=True)
Zomato_encoded.columns
```

higher the coreelation means higher the two variables related each other.
**The highest correlation is bettween reviews_list and city which is 0.74 And correlation between name and address which is 0.60**

**Rating and table  booking**

```python
plt.rcParams['figure.figsize']=(20,10)
Y=pd.crosstab(Zomato['rate'],Zomato['book_table'])
y=Y.div(Y.sum(1).astype(float),axis=0)
y.plot(kind='bar',stacked=True,color=['#849bad','#03fcca'])
plt.title("Rate  Vs  Table_booking",fontweight="bold",fontsize=30)
plt.legend(loc="upper right")
plt.ylabel("Table booking",fontsize=10,fontweight='bold')
plt.xlabel("Rate",fontsize=10,fontweight='bold')
plt.show()
```
