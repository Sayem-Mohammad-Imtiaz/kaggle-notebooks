```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
import plotly.figure_factory as ff
```

```python
netflix_filepath = "/kaggle/input/netflix-original-films-imdb-scores/NetflixOriginals.csv"
netflix_data = pd.read_csv(netflix_filepath, encoding="ISO-8859-1")
```

```python
t = np.linspace(-1, 1.2, 2000)
x = (t**3) + (0.3 * np.random.randn(2000))
y = (t**6) + (0.3 * np.random.randn(2000))

colorscale = ['#3A405A', '#AEC5EB', '#E9AFA3', '#F9DEC9', (0.98,0.98,0.98)]

fig = ff.create_2d_density(
    x=netflix_data['IMDB Score'], y=netflix_data['Runtime'], colorscale=colorscale,
    hist_color='rgb(255, 237, 222)', point_size=3, title="Title"
)

py.iplot(fig, filename ='netflix_data')
```
