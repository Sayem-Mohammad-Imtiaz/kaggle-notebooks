The main focus of this kernel is to analyse survey results of people who are either full time or part time employed in the industry. Studying about employed people would help others who are looking to get employed in the industry.

Also I would be using subset of columns to understand various trends among this particular group of people.
Those columns are :
**['Country','GenderSelect','Age','MLToolNextYearSelect','CurrentJobTitleSelect','LanguageRecommendationSelect',
'LearningPlatformSelect','JobSkillImportanceDegree','JobSkillImportanceKaggleRanking',
'JobSkillImportanceMOOC','FormalEducation','MajorSelect']**

In the first step, we read and filter the data according to our current focus of study.


```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

dataset = pd.read_csv('../input/multipleChoiceResponses.csv',encoding='ISO-8859-1')
columns_needed = ['Country','GenderSelect','Age','MLToolNextYearSelect','CurrentJobTitleSelect','LanguageRecommendationSelect','LearningPlatformSelect','JobSkillImportanceDegree','JobSkillImportanceKaggleRanking','JobSkillImportanceMOOC','FormalEducation','MajorSelect']
employee_dataset = pd.DataFrame(dataset[(dataset['EmploymentStatus']=='Employed full-time')|(dataset['EmploymentStatus']=='Employed part-time')],columns=columns_needed)
# Any results you write to the current directory are saved as output.
employee_dataset.head()
```

We will start with some basic statisitics related country and age.

```python
# Top 10 countries ranked according to number of respondent
print(employee_dataset['Country'].value_counts()[0:10]);
```

Degree distribution among the responders. Most of the responders who are employed have master's degree, followed by bachelor's and doctoral. 


```python
# age distribution of respondents 
print("Age stats")
print("Mean :",employee_dataset['Age'].mean())
print("Median :",employee_dataset['Age'].median())
print("Mode:",employee_dataset['Age'].value_counts().index[0])
ax=employee_dataset['Age'].hist(bins=100,color='blue',edgecolor='black',figsize=(10,10));
ax.grid(False);
ax.set_xlabel('Age');
ax.set_ylabel('Count');
```

Most of respondents have their age in the range of [20,40]. 

```python
ax=employee_dataset['FormalEducation'].value_counts().plot.pie(figsize=(10,10),autopct='%.2f%%',shadow=True)
ax.set_ylabel('');
```

Now lets us look at the what degree is held by most number of people having a given job titles. For this, we first do a group by **CurrentJobTitleSelect** operation on our **employee_details ** dataset and then calculate the mode for degree column
The table below has the degree held by most responders for a given job title along with percentage.

```python
employee_dataset.groupby(['CurrentJobTitleSelect']).agg(lambda x: (x.value_counts().index[0],x.value_counts().iloc[0]*100.0/x.value_counts().sum()))[['FormalEducation']]
```

57% of Scientist/Reseachers have a Doctoral degree. 46% of DBA/Database Engineers and Programmers have Bachelor's degree, while that number is 45% of Software Developer/Software Engineer. 

Now we will look what is the major field of study for different groups below.

```python
employee_dataset.groupby(['CurrentJobTitleSelect']).agg(lambda x: (x.value_counts().index[0],x.value_counts().iloc[0]*100.0/x.value_counts().sum()))[['MajorSelect']]
```

Well! Did we not expect that? 66% of Statistician, 24% of Data Analyst, 24% Data Scientist, 22.5% Operations Research Practitioners have done their majors in Mathematics or statistics.  While 69% of Computer Scientists, 55% of Software Developers, 48% of Programmer have majors in Computer Science.





```python
ax=employee_dataset['LanguageRecommendationSelect'].value_counts(sort=True).plot.pie(figsize=(10,10),autopct='%.2f%%',shadow=True,startangle=45)
ax.set_ylabel('');
```

Python is the clear winner with 62% of respondents recommending it. It is followed by R (25%) and SQL(4%). Do we find anything interesting when we look at recommendation according to job titles? Let us find out



```python
employee_dataset.groupby(['CurrentJobTitleSelect']).agg(lambda x: (x.value_counts().index[0],x.value_counts().iloc[0]*100.0/x.value_counts().sum()))[['LanguageRecommendationSelect']]
```

Except for Operations Research Practitioner and Statisiticians who recommend R, majority of people belonging to other job titles recommend python
The other interesting thing one would look out for is , what currently employed people excited about learning in the next year. 

```python
ax=employee_dataset['MLToolNextYearSelect'].value_counts(sort=True).plot.pie(figsize=(10,10),autopct='%.2f%%',shadow=True,startangle=45)
ax.set_ylabel('');
```

24.77% of employed respondents are excited learning TensorFlow, whil1 14.77% for python, 8.10% for R and 7.75% for Spark/Mlib. We can see that there are lots of other tools about which people are excited to learn. 

Lets look at group wise most interested tool that people want to learn

```python
employee_dataset.groupby(['CurrentJobTitleSelect']).agg(lambda x: (x.value_counts().index[0],x.value_counts().iloc[0]*100.0/x.value_counts().sum()))[['MLToolNextYearSelect']]
```

More than 20% of Business Analysts, Statisticians, Data Analyst and nearly 19% of Data miners, Predictive Modeler and 13% of Operations Research Practitioners want to are excited about learning python. On the other hand 31% of Software Developers, 30% of Data Scientists, 28% of Computer Scientists, 34% of ML engineers, 27% of Programmers and 26% of Engineers are excited about learning TensorFlow.
