***Question 4) use the function boxplot in the seaborn library to produce a plot that can be used to determine whether houses with a waterfront view or without a waterfront view have more price outliers.***

```python
with_view = king_sales.loc[king_sales['waterfront'] > 0]
with_view
```

```python
without_view = king_sales.loc[king_sales['waterfront'] == 0]
without_view
```

```python
view = pd.DataFrame(data = np.random.random(size=(4,3)), columns = ['with_view', 'without_view', 'price'])

sns.boxplot(x="variable", y="value", data=pd.melt(view))

plt.show()
```
