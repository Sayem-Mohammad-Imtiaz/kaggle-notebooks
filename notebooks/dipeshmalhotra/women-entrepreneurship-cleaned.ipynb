The data is in ; seperated format. Let's pre process the data to get started.

# Data Preprocessing

```python
new=df['No;Country;Level of development;European Union Membership;Currency;Women Entrepreneurship Index;Entrepreneurship Index;Inflation rate;Female Labor Force Participation Rate']
new=new.str.split(";")
```

```python
cols=df.columns.values
s= "".join(cols)
cols=s.split(";")
```

```python
new_df=pd.DataFrame()
for row in new:
    S_row=pd.Series(row)
    new_df=new_df.append(S_row,ignore_index=True)
```

```python
new_df.columns=cols
new_df.drop('No',axis=1,inplace=True)
new_df.head()
```

```python
new_df.info()
```

```python
num_cols=['Women Entrepreneurship Index','Entrepreneurship Index',        # Numerical columns    
  'Inflation rate' ,'Female Labor Force Participation Rate']
```

```python
new_df[num_cols]=new_df[num_cols].astype(float)
```

**Renaming long columns to make your job easier :D**

```python
new_df.rename(columns={'Level of development':'Developed',
                       'European Union Membership':'EU_Member',
                       'Women Entrepreneurship Index':'Women_Ent_Idx',
                       'Entrepreneurship Index':'Ent_Idx',
                       'Inflation rate':'Inflation',
                       'Female Labor Force Participation Rate':'Female_labor_prate'},
              inplace=True)
```

```python
new_df.info()
```

```python
px.scatter(new_df,x='Ent_Idx',y='Women_Ent_Idx',color='EU_Member')
```

**This plot tell us:**

A country with higher entrepreneurship index has higher chances of having a higher women entrepreneurship index
