```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
#loading and checking the data set 
df=pd.read_csv('../input/breast-cancer-wisconsin-data/data.csv',index_col='id')
df.head()
# df.tail()
```

```python
#setting a random seed 
import random
rs=7
random.seed(rs)
```

```python
#describe the dataset 
df.describe()
#info method on dataset
df.info()
```

```python
#checking for null values and dropping them 
# df.isnull().sum()
# we see that last col is null so we drop it  
df.dropna(inplace=True,axis=1)
df.isnull().sum()
```

```python
#the diagnosis column is the target column with its dtype= object.
#lets see the number of unique observations in that column 
# df.diagnosis.nunique()

#checking number of examples in diagnosis :

sns.countplot(df['diagnosis'])
```

```python
#hence the data shows the that there are 357 instances of begign tumors 
#and 212 intances of malignant tumor 
#lets encode the malignant as 1's and begign as 0's
df['diagnosis']=df['diagnosis'].map({'M':1, 'B':0})
df.diagnosis
```

```python
#lets see if there are any categorical columns in the dataset 
cat_cols=[col for col in df.columns if df[col].dtype=='object']
cat_cols
#there are no categorical columns in the dataset 
```

```python
#lets plot the correlation heatmap
correlation_matrix=df.corr()
# correlation_matrix
sns.heatmap(correlation_matrix)
plt.title('Correlation matrix')
plt.show()



#lets drop columns with correlation <.1 with the diagnosis
cor_cols=[col for col in df.columns if abs(df[col].corr(df['diagnosis']))>.1]
#list of columns with correlation greater than .1 with the target.
cor_cols
```

```python
#the data is clean and does not have categorical variables,so that we can proceed
# to train the model 
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier 
from sklearn.linear_model import LinearRegression,LogisticRegression 
from sklearn.neighbors import KNeighborsClassifier 
import xgboost as xgb


#split data into training and testing sets
X=df[cor_cols].drop('diagnosis',axis=1)
y=df.diagnosis

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.3,shuffle=True,random_state=rs)

model1=LinearRegression()
model2=LogisticRegression(random_state=rs)
model3=KNeighborsClassifier()
model4=RandomForestClassifier(random_state=rs)
model5=xgb.XGBClassifier(random_state=rs,n_estimators=500
                        )




#Training and testing the models

model1.fit(X_train,y_train)
a=model1.score(X_test,y_test)
print('the score with {} is {}'.format(model1,a))

model2.fit(X_train,y_train)
a=model2.score(X_test,y_test)
print('the score with {} is {}'.format(model2,a))

model3.fit(X_train,y_train)
a=model3.score(X_test,y_test)
print('the score with {} is {}'.format(model3,a))

model4.fit(X_train,y_train)
a=model4.score(X_test,y_test)
print('the score with {} is {}'.format(model4,a))

model5.fit(X_train,y_train)
a=model5.score(X_test,y_test)
print('the score with {} is {}'.format(model5,a))
```

```python
#let us plot the confusion matrix for Random Forest Classifier and XGBClassifier
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(model4,X_test,y_test)
plt.title('Confusion Matrix for RandomForestClassifier')
plt.show()

plot_confusion_matrix(model4,X_test,y_test)
plt.title('Confusion Matrix for XgbClassifier')
plt.show()
```
