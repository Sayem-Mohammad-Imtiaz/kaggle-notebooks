# Scraping web for some financial insights : 

```python
all_tables=pd.read_html('https://prsindia.org/budgets/parliament/demand-for-grants-2020-21-analysis-health-and-family-welfare#_ednref11')
```

```python
df1= pd.DataFrame(all_tables[8])
```

```python
df1.columns=list(df1.iloc[0,:])
df1=df1.iloc[1:-1,:]
```

```python
df1.sort_values('State',inplace=True)
df1.reset_index(drop=True,inplace=True)
```

```python
df1['Total']= df1['Average health expenditure (rural)'].apply(int)+df1['Average health expenditure (urban)'].apply(int)
```

```python
df1.head()
```

```python
col= df1['Total']
df['Health expenditure']= col
```

```python
df.head()
```

```python
plt.figure(figsize=(15,20))
sn.barplot(data=df,y='State/UTs',x='Health expenditure')
plt.suptitle('Average Health Expenditure of States',fontsize=28)
plt.show()
```

```python
LS = df['Total Cases'].min() 
LE = df['Total Cases'].max()  
```

**Since health expenditure and total cases ideally should be inversely proportional to one another, as greater health expenditure means better infrastructure, I drew a line with a negative slope to roughly approximate the curve. Two states/UT have performed poorly, with Maharashtra being the worst performer.**

```python
g=sn.relplot(data=df,y= 'Total Cases',x='Health expenditure',hue='State/UTs',s=200,height=8,aspect=2)
plt.suptitle('Health Expenditure vs Total Cases',fontsize=20)
plt.plot([0, 70000], [LE, 0], 'k-', color = 'r')
plt.show()
```

**As per the data, reported deaths in Maharashtra have also been the highest when considering the average amount spent by the sate on healthcare**

```python
g=sn.relplot(data=df,y= 'Deaths',x='Health expenditure',hue='State/UTs',s=200,height=10,aspect=1.5)
plt.suptitle('Health Expenditure vs Deaths',fontsize=20)
plt.show()
```

```python
plt.figure(figsize= (10,15))
sn.barplot(data=df, x ='Discharge Ratio (%)',y='State/UTs')
plt.show()
```

**Almost every state has the same Cases-discharged ratio**
