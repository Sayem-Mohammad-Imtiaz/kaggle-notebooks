# New T17:

```python
#Lert's make a new 'region' feature for T17 dataset
Region = []
T17Coun = T17['Country']

for i in T17Coun:
    if i in Western_Europe:
        Region.append('Western Europe')
        
    elif i in Sub_Saharan_Africa:
         Region.append('Sub-Saharan Africa')
            
    elif i in Central_and_Eastern_Europe:
         Region.append('Central and Eastern Europe')
            
    elif i in Latin_America_and_Caribbean:
         Region.append('Latin_America_and_Caribbean')
            
    elif i in Middle_East_and_Northern_Africa:
         Region.append('Middle East and Northern Africa')
            
    elif i in Southeastern_Asia:
         Region.append('Southeastern_Asia')
            
    elif i in Eastern_Asia:
         Region.append('Eastern Asia')
            
    elif i in Australia_and_New_Zealand:
         Region.append('Australia and New Zealand')
            
    elif i in North_America:
         Region.append('North America')
            
    elif i in Southern_Asia:
         Region.append('Southern Asia')
            
    else:
         Region.append('Others')
len(Region)
```
