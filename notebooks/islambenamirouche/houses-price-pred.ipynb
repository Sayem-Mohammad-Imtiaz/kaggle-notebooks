# 1. **importing dataset**

```python
dataset = pd.read_csv("/kaggle/input/real-estate-price-prediction/Real estate.csv")
dataset.head()
```

# 2. dataset info

```python
#let's see if our dataset contains null values
print(dataset.info())
```

# dependent and independent values

```python
#multivariate linear regression
#i will ignore the 2 first columns Number+Date for this moment and we will see
X=dataset.iloc[:,2:-1].values
y=dataset.iloc[:,-1].values
print(X[:5,:])
```

# train test splitting

```python
#let's split our data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1,random_state = 0)
print(X_train.shape)
```

# training our model

```python
#now let's create our model
from sklearn.linear_model import LinearRegression
lin_regressor=LinearRegression()
lin_regressor.fit(X_train,y_train)
```

```python
y_pred=lin_regressor.predict(X_test)
```

# comparing predicted data and real data
> there are some cases where there is large difference between predicted value and the real one example:  45.739271 & 26.5
but generaly the model is accpetable

```python
#now let's compare if it works perfectly
con=np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),axis=1)
compare_df=pd.DataFrame(data=con,columns=["predicted_value","real_value"])
compare_df
```
