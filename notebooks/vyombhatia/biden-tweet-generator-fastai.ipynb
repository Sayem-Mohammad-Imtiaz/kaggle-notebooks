# Importing Stuff

```python
import pandas as pd
import numpy as np

# Sci-Kit Learn, FastAI and Torch:
from fastai.text.all import *
import torch

# 
from sklearn.metrics import accuracy_score, precision_recall_fscore_support
```

# Grabbing Data and Models

Lets import the data and take a look at it:

```python
path = "../input/us-election-2020-tweets/hashtag_joebiden.csv"
data = pd.read_csv(path)
data = data['tweet']
data.head()
```

# Working with 60,000 lines of text

# Preprocessing the Data:

```python
train = data.loc[:90000]

train = pd.DataFrame(train)
```

Setting up the tokenizer:

```python
dls = TextDataLoaders.from_df(train, text_col = 'tweet', is_lm = True, seq_len = 50)
```

```python
dls.show_batch(max_n=3)
```

# Training the Model

Using FastAI's language_model_learner module:

```python
learn = language_model_learner(dls, AWD_LSTM, drop_multi = 0.5, metrics = accuracy)
```

Finding a suitable learning rate:

```python
learn.lr_find()
```

It seems like a suitable learning rate lies around 1e-2, because the slope is the steepest there.

Lets train the model for a few epochs now:

```python
learn.fit_one_cycle(5, 1e-2)
```

```python
learn.predict('biden sucks so much, he will raise my taxes.',10, temperature = 1)
```

# Working with 200,000 lines of text

## Preprocessing the Data (Again)

```python
train = data.loc[:200000]

train = pd.DataFrame(train)
```

Setting up the dataloader:

```python
dls = TextDataLoaders.from_df(train, text_col = 'tweet', is_lm = True, seq_len = 50)
```

## Training the Model (Again)

```python
learn = language_model_learner(dls, AWD_LSTM, drop_multi = 0.5, metrics = accuracy)
```

Lets train the model for a few epochs now:

```python
learn.fit_one_cycle(1, 1e-2)
```

```python
learn.predict('I dont like biden at all he sucks. ', 40,temperature = 1)
```
