# Preprocessing Texts

## Preprocessing Function

```python
import re

def preprocess(text):
  o = re.sub(r'\\', ' ', text)
  k = re.sub(r'\n', ' ', o)
  l = re.sub(r'&', '', k)
  m = re.sub(r'RT ', '', l)
  n = re.sub(r'~', '', m)
  o = re.sub(r'#', '', n)
  p = re.sub(r'!+', '', o)
  q = re.sub(r'(http[s]?\S+)|(\w+\.[A-Za-z]{2,4}\S*)', 'link', p)
  r = re.sub(r'[*]', '', q)
  s = re.sub(r'[@]\w+', 'user', r)
  t = re.sub(r'[:|;]', '', s)
  u = re.sub(r'[\\x]\W+|\d', '', t)
  v = re.sub(r'[\\x]\W+|\d', '', u)
  w = re.sub(r'[\\x]\w+', '', v)
  x = re.sub(r'    ', ' ', w)
  y = re.sub(r'   ', ' ', x)
  z = re.sub(r'  ', ' ', y)
  ProcessedTweet = z.lower()
  return ProcessedTweet
```

### Preprocessing the texts finally:

```python
for i in tqdm(train.index):
    train.loc[i] = preprocess(train.loc[i])
```

```python
dls = TextDataLoaders.from_df(train, path="./", text_col='content', valid_pct=0.3, is_lm=True)
```

# Model

It's a simplpe language model which generates sentences. 

```python
model = language_model_learner(dls, AWD_LSTM)
```

Training the model:

```python
model.fit_one_cycle(5)
```
