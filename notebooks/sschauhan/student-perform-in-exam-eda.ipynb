```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
import plotly.graph_objects as go
import plotly.express as px
import plotly.figure_factory as ff
```

## import the data

```python
df= pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
```

## EDA

```python
df['gender'].value_counts()
```

```python
sns.set_style('whitegrid')
sns.countplot(x=df['gender'], data=df)
plt.xlabel('gender')
plt.ylabel('frequency')
plt.title('Gender Count Plot')
```

```python
df['race/ethnicity'].value_counts()
```

```python
sns.set_style('darkgrid')
sns.barplot(x=df['race/ethnicity'].value_counts().index,
              y=df['race/ethnicity'].value_counts().values, data=df)
plt.xlabel('Race/Ethinicity')
plt.ylabel('No of Frequency')
plt.title('Race/Ethinicity Count Plot')
```

```python
df['parental level of education'].value_counts()
```

```python
plt.figure(figsize=(10,6))
sns.set_style('darkgrid')
sns.barplot(x='parental level of education',y='writing score', hue='gender', data=df)
plt.xticks(rotation=-40)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.set_style('whitegrid')
sns.barplot(x='parental level of education',y='reading score', hue='gender', data=df)
plt.xticks(rotation=-40)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.catplot(x="gender", y="math score",
                 hue="parental level of education",
                 data=df, kind="bar")
plt.title('for Parental Level Of Education Gender & Math Score')
```

```python
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.countplot(x='gender', hue='test preparation course',data=df )
plt.legend(loc='upper center')
```

```python
plt.figure(figsize=(8,5))
sns.set_style('dark')
sns.countplot(x='parental level of education', hue='test preparation course',data=df )
plt.legend(loc='upper right')
plt.xticks(rotation=-35)
```

```python
df['lunch'].value_counts()
```

```python
sns.countplot(x='lunch',data=df )
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='parental level of education', hue='lunch', data=df)
plt.xlabel('Parental Level of Education')
plt.legend(loc='upper right')
plt.xticks(rotation=-40)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='race/ethnicity', hue='lunch', data=df)
plt.xlabel('Race Ethnicity')
plt.legend(loc='upper right')
plt.xticks(rotation=-40)
```
