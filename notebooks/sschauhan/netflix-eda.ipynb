# Import the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Import Data

```python
netflix_data_df= pd.read_csv('../input/netflix-shows/netflix_titles.csv')
netflix_data_df.head()
```

# Check the Data

```python
netflix_data_df.info()
```

```python
netflix_data_df.nunique()
```

# Find the null values

```python
netflix_data_df.isnull().sum().sum()
```

# null values heatmap

```python
sns.heatmap(netflix_data_df.isnull(), cbar=False)
plt.title('Null Values Heatmap')
plt.show()
```

```python
netflix_data_df.isnull().sum()
```

# fix the null values

```python
netflix_data_df['director'].fillna('No Director', inplace=True)
netflix_data_df['cast'].fillna('No Cast', inplace=True)
netflix_data_df['country'].fillna('Country Unavailable', inplace=True)
netflix_data_df.dropna(subset=['date_added', 'rating'], inplace=True)
```

```python
netflix_data_df.isnull().sum()
```

## Check the data again after fixing the null values

```python
netflix_movies_df= netflix_data_df[netflix_data_df['type']=='Movie'].copy()
netflix_movies_df.head()    
```

```python
netflix_shows_df=netflix_data_df[netflix_data_df['type']=='TV Show'].copy()
netflix_shows_df.head()
```

# Data Preparation

```python
netflix_movies_df.duration=netflix_movies_df.duration.str.replace('min','').astype(int)
netflix_shows_df.rename(columns={'duration':'seasons'}, inplace=True)
netflix_shows_df.replace({'seasons':{'1 Season':'1 Seasons'}}, inplace=True)
netflix_shows_df.seasons= netflix_shows_df.seasons.str.replace('Seasons','').astype(int)
```

# EDA

```python
netflix_data_df.head()
```

# Types of Content on Netflix

```python
plt.figure(figsize=(7,5))
sns.countplot(netflix_data_df.type);
plt.xlabel('Type (Movie/ TV Shows)')
plt.ylabel('Total Count')
plt.title('Count of Movies and TV Shows')
plt.show()
```

# in percentage

```python
plt.figure(figsize=(12,6))
plt.title('% of Netflix Titles That are either Movies or TV Shows')
plt.pie(netflix_data_df.type.value_counts(),labels=netflix_data_df.type.value_counts().index, autopct='%1.1f%%', startangle=180);
plt.show()
```

# Netflix ratings

```python
plt.figure(figsize=(15,7))
sns.countplot(netflix_data_df.rating, hue=netflix_data_df.type);
plt.title("Rating for Movies and TV Shows")
plt.xlabel('Rating')
plt.ylabel('Total Count')
plt.show()
```

# netflix movies and shows rating

```python
plt.subplots(1,2,figsize=(18,6))
plt.subplot(121)
sns.countplot(netflix_movies_df.rating)
plt.title('Movies Rating')
plt.xlabel('Rating')
plt.ylabel('Value Counts')
plt.subplot(122)
sns.countplot(netflix_shows_df.rating)
plt.title('TV Shows Rating')
```

## content added each year

```python
netflix_data_df['year_added']=pd.DatetimeIndex(netflix_data_df['date_added']).year
netflix_movies_df['year_added']=pd.DatetimeIndex(netflix_movies_df['date_added']).year
netflix_shows_df['year_added']=pd.DatetimeIndex(netflix_shows_df['date_added']).year
netflix_data_df['month_added']=pd.DatetimeIndex(netflix_data_df['date_added']).month
netflix_movies_df['month_added']=pd.DatetimeIndex(netflix_movies_df['date_added']).month
netflix_shows_df['month_added']=pd.DatetimeIndex(netflix_shows_df['date_added']).month
```

```python
netflix_year=netflix_data_df['year_added'].value_counts()
netflix_year
```

# Netflix Movies Duration and TV Shows Seasons

```python
plt.subplots(1,2, figsize=(18,6))
plt.subplot(121)
sns.distplot(netflix_movies_df.duration)
plt.title('Duration for Netflix Movies')
plt.xlabel('Duration (Min)')
plt.ylabel('% of all Netflix Movies')
plt.subplot(122)
sns.countplot(netflix_shows_df.seasons)
plt.title('Seasons of Netflix TV Shows')
plt.xlabel('Seasons')
plt.ylabel('Count')
plt.show()
```

# Countries with most content

```python
country= netflix_data_df[netflix_data_df['country']!= 'Country Unavailable'].set_index('title').country.str.split(',', expand=True).stack()
plt.figure(figsize=(7,9))
sns.countplot(y=country, order=country.value_counts().index[:20])
plt.title('Top 20 Country on Netflix')
plt.xlabel('Titles')
plt.ylabel('Country')
plt.show()
```

# popular genre

```python
plt.figure(figsize=(7,9))
sns.countplot(y=netflix_data_df['listed_in'], order=netflix_data_df['listed_in'].value_counts().index[:15])
plt.title('Top 15 Genres on Netflix')
plt.xlabel('Titles')
plt.ylabel('Genres')
plt.show()
```

## What are the top 10 Directors on Netflix?

# Top 10 Directors

```python
director=netflix_data_df[netflix_data_df.director!='No Director'].set_index('title').director.str.split(',', expand=True).stack()
sns.countplot(y=director, order=director.value_counts().index[:10])
plt.title('Top 10 Netflix Directors')
```

## what are the top 10 Actors ?

```python
cast= netflix_data_df[netflix_data_df['cast']!='No Cast'].set_index('title').cast.str.split(',', expand=True).stack()
sns.countplot(y= cast, order=cast.value_counts().index[:10])
plt.show()
```
