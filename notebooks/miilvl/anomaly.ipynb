```python
def KMeans_metric(X, N = 3):
    sizeX = len(X[0])
    N_Closest = np.zeros(sizeX)
    for i in range(sizeX):
        for ns in range(N):
            prevSmallest = 0;
            for j in range(sizeX):
                length = (X[0][1] - X[0][j])**2 + (X[1][i] - X[1][j])**2 + (X[2][i] - X[2][j])**2
                if (N_Closest[i] == prevSmallest) or ((N_Closest[i] > length) and (i != j) and (length != prevSmallest)):
                    N_Closest[i] = length
            prevSmallest = N_Closest[i]
    me = sum(N_Closest)/sizeX
    ds = (sum((sum(N_Closest**2)/sizeX) - me**2))**0.5
    res = np.zeros(sizeX)
    print(me, ds)
    for i in range (sizeX):
        if abs(N_Closest[i] - me) < ds:
            res[i] = 1
    return res
    
```

```python
res = KMeans_metric(dset)
res
```

```python
plt.scatter(data['Age'], data['Spending Score (1-100)'], c=res, cmap = 'hot')
plt.xlabel('Age')
plt.ylabel('Score')
plt.show()
```
