```python
books = pd.read_csv('/kaggle/input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
books.head()
```

```python
books.info()
```

## Genre

```python
books.Genre.value_counts().plot.pie(autopct='%0.2f%%',title='Genre popularity')
plt.show()
```

Non fiction books are sold more than fiction books.

```python
plt.figure(figsize=[12, 8])
sns.countplot(data = books, x = 'Year', hue = 'Genre')
plt.show()
```

The popularity of fiction & non fiction has been fluctuating over the years. However, in the year 2014, fiction's popularity surpassed non fiction. 

## Price Trends

```python
books.groupby('Year').Price.mean().plot.bar(figsize=[10,6])
plt.title('Price trend over the years',fontsize =14)
plt.show()
```

Over the years the average price of novels has dropped. This can be attributed to the rising popularity of e-books which are available at cheaper prices.

```python
year_genre = pd.pivot_table(data = books, columns = 'Year',index = 'Genre', values = 'Price',aggfunc=np.mean)
plt.figure(figsize=[15,6])
sns.heatmap(year_genre,cmap = 'Greens',center = 10)
plt.yticks(rotation = 0)
plt.title('Price trend of the different Genre over the years',fontsize = 14)
plt.show()
```

Non fiction has been particularly costly during the year 2014. This can be attributed to the fact that fiction gained more popularity than non fiction in 2014.

## Best Selling Authors

```python
plt.figure(figsize=[12,8])
sns.countplot(data = books, x = 'Author', order = books.Author.value_counts().iloc[:10].index)
plt.xticks(rotation = 80,weight='bold',size = 10)
plt.title('Best selling authors over the year')
plt.show()
```

## Reviews

```python
books.sort_values(by='Reviews',ascending = False,inplace = True)
books.head(20)
```

Multiple books have been in the top 50 bestsellers for more than one year. 

```python
max_reviews = books.drop_duplicates(['Name'])
max_reviews.head(20)
```

```python
plt.figure(figsize = [20,8])
sns.barplot(data = max_reviews.head(20),x = 'Name',y='Reviews')
plt.xticks(rotation = 90)
plt.title('Books with maximum reviews on Amazon',fontsize=14,fontweight='bold')
plt.show()
```

```python
year_genre = pd.pivot_table(data = max_reviews, columns = 'Year',index = 'Genre', values = 'Reviews',aggfunc=np.mean)
plt.figure(figsize=[15,6])
sns.heatmap(year_genre,cmap = 'Greens',center = 4000)
plt.yticks(rotation = 0)
plt.title('Reviews of the different Genre over the years',fontsize = 14)
plt.show()
```

Fiction got maximum reviews in 2015.

## User Ratings

```python
books.sort_values(by = 'User Rating',ascending = False,inplace = True)
books.drop_duplicates(['Name'],inplace = True)
books.head(20)
```

75% of the books in Top 20 are fiction.

```python
year_genre = pd.pivot_table(data = books, columns = 'Year',index = 'Genre', values = 'User Rating',aggfunc=np.mean)
plt.figure(figsize=[15,6])
sns.heatmap(year_genre,cmap='Greens')
plt.yticks(rotation = 0)
plt.title('User Rating trend of the different Genre over the years',fontsize = 14)
plt.show()
```

Fiction has been highly rated for the year 2019.

Though non fiction is sold more than fiction, fiction is more review and better rated than non fiction.
