```python
d = pd.read_csv('/kaggle/input/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv')
print(d.head())
```

```python
df = d.sort_values(by='Death Ratio (%)', ascending=False).head()
states = df['State/UTs']
ratio = df['Death Ratio (%)']
plt.barh(states, ratio)
plt.xlabel('Death Ratio (%)')
plt.ylabel('State')
plt.title('State with  more death ratio in India')
plt.show()
```
