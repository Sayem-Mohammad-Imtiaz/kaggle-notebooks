
**VISUAL  DATA ANALYSIS**
We have to know some basic definitions
Outliers: the value that is considerably higher or lower from rest of the data(datanın geri kalanından hatırı sayılır derece ayrık olan veri, aykırı olan veri)

What is quantile?
1,4,5,6,8,9,11,12,13,14,15,16,17 this is sample list.

*The median is the number that is in middle of the sequence. In this case it would be 11.
*The lower quartile is the median in between the smallest number and the median i.e. in between 1 and 11, which is 6.    Lower quartile=Q1=%25
*The upper quartile, you find the median between the median and the largest number i.e. between 11 and 17, which will be 14 according to the question above. Upper quartile ise Q3=%75 
*Lets say value at 75% is Q3 and value at 25% is Q1.
*Outlier are smaller than Q1 - 1.5(Q3-Q1) and bigger than Q3 + 1.5(Q3-Q1). (Q3-Q1) = IQR
*We will use describe() method. Describe method includes
*count: number of entries
*mean: average of entries
*std: standart deviation
*min: minimum entry
*25%: first quantile
*50%: median or second quantile
*75%: third quantile
*max: maximum entry

TIDY DATA
We tidy data with melt()
Data melting is the way of the make  features as a variable in the row and features  values as a value in the row again.
For example:

```python
# Firstly I create new data from dataFrame data to explain melt more easily.
data_new = dataFrame.head()    # I only take 5 rows into new data
data_new
```

```python
# melting
# id_vars = what we do not wish to melt
# value_vars = what we want to melt
# column name  variable ve value is default
melted = pd.melt(frame=data_new,id_vars = 'Glucose', value_vars= ['BMI','Age'])
melted
```

```python
# Reverse of melting.
# Index is Glucose
# after reverse melting i want to next feature name is melted columns variable
# Finally values in columns are value
melted.pivot(index = 'Glucose', columns = 'variable',values='value')
```




CONCATENATING DATA

We can concatenate two dataframe
there are two type concatenating 


adds dataframes in row ----this is vertical

adds dataframes in column ---this is horizontal 

this method part of pandas library



```python
data1 = dataFrame.head()
data2= dataFrame.tail()
conc_data_row = pd.concat([data1,data2],axis =0,ignore_index =True) # axis = 0 : adds dataframes in row
conc_data_row

#ignore index:True giving new indexing number after the concat.
```

```python
#column concatenating, horizontal 
data1 = dataFrame['BloodPressure'].head()
data2= dataFrame['SkinThickness'].head()
conc_data_col = pd.concat([data1,data2],axis =1) # axis = 1 : adds dataframes in column
conc_data_col
```
