```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../input/ibm-hr-analytics-attrition-dataset/WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df['Attrition']=df.Attrition.replace({'Yes':1,'No':0})
plt.figure(figsize=(16,9))
sns.barplot(x=df['Attrition'].value_counts().index,y=df['Attrition'].value_counts().values,palette='crest')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
```

```python
plt.figure(figsize=(16,9))
sns.boxplot(x='MonthlyIncome',y='Gender',data=df,palette='crest',orient='h')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Monthly Income',fontsize=14)
plt.ylabel('Gender',fontsize=14)
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x=df.groupby('Gender')['MonthlyIncome'].mean().index,y=df.groupby('Gender')['MonthlyIncome'].mean().values,palette='crest')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Gender',fontsize=14)
plt.ylabel('Average Monthly Income',fontsize=14)
```

```python
plt.figure(figsize=(16,9))
sns.countplot(x='Department',hue='Gender',data=df,palette='crest')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Department',fontsize=14)
plt.ylabel('Number of Employees',fontsize=14)
```

```python
fig, axes=plt.subplots(2,1,figsize=(16,9),sharex=False,sharey=False)
df['Education']=df.Education.replace({1:'Below College',2:'College',3:'Bachelor',4:'Master',5:'Doctor'})
sns.countplot(x='Education',hue='Attrition',data=df,palette='crest',ax=axes[0])
sns.barplot(x='Education',y='Attrition',data=df,palette='crest',ax=axes[1],ci=None)
axes[0].set_xlabel('Education',fontsize=14)
axes[0].set_ylabel('Count',fontsize=14)
axes[1].set_xlabel('Education',fontsize=14)
axes[1].set_ylabel('Attrition Rate',fontsize=14)
```

```python
plt.figure(figsize=(16,9))
sns.barplot(x='Department',y='MonthlyIncome',hue='Attrition',data=df,ci=None,palette='crest')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Department',fontsize=14)
plt.ylabel('Average Monthly Income',fontsize=14)
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x='PercentSalaryHike',y='MonthlyIncome',hue='Attrition',data=df)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Percent Salary Increase',fontsize=14)
plt.ylabel('Monthly Income',fontsize=14)
```

```python
fig, axes=plt.subplots(2,1,figsize=(16,9),sharex=False,sharey=False)
sns.countplot(x='OverTime',hue='Attrition',data=df,palette='crest',ax=axes[0])
sns.barplot(x='OverTime',y='Attrition',data=df,palette='crest',ci=None,ax=axes[1])
axes[0].set_xlabel('OverTime',fontsize=14)
axes[0].set_ylabel('Count',fontsize=14)
axes[1].set_xlabel('OverTime',fontsize=14)
axes[1].set_ylabel('Attrition Rate',fontsize=14)
```

```python
plt.figure(figsize=(16,9))
sns.kdeplot(x='YearsAtCompany',y='JobSatisfaction',data=df,shade=True,cmap='YlOrBr')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('Years At Company',fontsize=14)
plt.ylabel('Job Satisfaction',fontsize=14)
```
