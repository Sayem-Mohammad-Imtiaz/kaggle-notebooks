```python
def replace(experience):
    if experience == '>20':
        return 21
    elif experience == '<1':
        return 0

    else:
        return experience
```

```python
df_train['experience'] = df_train['experience'].map(replace)
```

```python
def replace(last_new_job):
    if last_new_job == '>4':
        return 5
    elif last_new_job == 'never':
        return 0

    else:
        return last_new_job
```

```python
df_train['last_new_job'] = df_train['last_new_job'].map(replace)
```
