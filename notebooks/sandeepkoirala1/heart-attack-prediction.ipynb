1.1 Data Dictionary
age - Age of the patient

sex - Sex of the patient

cp - Chest pain type ~ 0 = Typical Angina, 1 = Atypical Angina, 2 = Non-anginal Pain, 3 = Asymptomatic

trtbps - Resting blood pressure (in mm Hg)

chol - Cholestoral in mg/dl fetched via BMI sensor

fbs - (fasting blood sugar > 120 mg/dl) ~ 1 = True, 0 = False

restecg - Resting electrocardiographic results ~ 0 = Normal, 1 = ST-T wave normality, 2 = Left ventricular hypertrophy

thalachh - Maximum heart rate achieved

oldpeak - Previous peak

slp - Slope
the slope of the peak exercise ST segment
-- Value 1: upsloping
-- Value 2: flat
-- Value 3: downsloping

caa - Number of major vessels

thall - A blood disorder called thalassemia Value 0: NULL (dropped from the dataset previously
Value 1: fixed defect (no blood flow in some part of the heart)
Value 2: normal blood flow
Value 3: reversible defect (a blood flow is observed but it is not normal)

exng - Exercise induced angina ~ 1 = Yes, 0 = No

output - Target variable

### Label Encoding data

```python
#Creating a copy of the Dataset
df1=df
```

```python
df1 = pd.get_dummies(df1, columns = cat_cols, drop_first = True)
```

```python
df1.head()
```

```python
#Treating the Outliers in the data
def remove_outliers(col):
    Q1,Q3 = np.percentile(col,[25,75])
    IQR=Q3-Q1
    Lower_range=Q1-(IQR*1.5)
    Upper_range =Q3 + (IQR*1.5)
    return Lower_range, Upper_range
```

```python
for i in df1[con_cols]:
    Lr,Ur= remove_outliers(df1[i])
    df1[i]=np.where(df1[i]<Lr,Lr,df1[i])
    df1[i]=np.where(df1[i]>Ur,Ur,df1[i])
```

```python
for i in df1[con_cols]:
    j=1
    while(j<9):
        axes=plt.subplots(nrows=j,ncols=1)
        sns.boxplot(df1[i])
        j=j+1
        break
```

```python
#Outliers have been removed
```

```python
#Scaling the data
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df1[con_cols] = scaler.fit_transform(df1[con_cols])
print("The first 5 rows of data are")
df1.head()
```

### Logestic Regression

```python
X=df1.drop('output',axis=1)
Y=df1[['output']]
```

```python
from sklearn.model_selection  import train_test_split
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.30,stratify=Y,random_state=11)
```

```python
LR=LogisticRegression(solver='newton-cg',max_iter=10000)
LR_Model=LR.fit(X_train,Y_train)
```

```python
LR_train_score = LR_Model.score(X_train, Y_train)
print('Accuracy Score of training data is ',LR_train_score)
LR_test_score = LR_Model.score(X_test, Y_test)
print('Accuracy Score Test Data is ',LR_test_score)
```

```python
#Performance on training Data
```

```python
LR_Train_predict=LR_Model.predict(X_train)
LR_Test_predict=LR_Model.predict(X_test)
```

```python
LR_Train_prob=LR_Model.predict_proba(X_train)
LR_Train_prob=LR_Train_prob[:,1]
```

```python
LR_Test_prob=LR_Model.predict_proba(X_test)
LR_Test_prob=LR_Test_prob[:,1]
```

```python
Lr_train_auc=roc_auc_score(Y_train,LR_Train_prob)
print('Area under the curve of ROC curve in Training Data',Lr_train_auc)
Lr_test_auc=roc_auc_score(Y_test,LR_Test_prob)
print('Area under the curve of ROC curve in Testing Data',Lr_test_auc)
```

```python
#Roc curve training and testing data Logical Regression
Lr_train_fpr,Lr_train_tpr,thresholds=roc_curve(Y_train,LR_Train_prob)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(Lr_train_fpr, Lr_train_tpr,color='red')

Lr_test_fpr,Lr_test_tpr,thresholds=roc_curve(Y_test,LR_Test_prob)
plt.plot([0, 1], [0, 1], linestyle='--')
plt.plot(Lr_test_fpr, Lr_test_tpr,color='green')
```

```python
print('confussion matrix','\n',metrics.confusion_matrix(Y_train,LR_Train_predict))
print('Classification Report','\n',metrics.classification_report(Y_train,LR_Train_predict))
```

```python
LR_train_metrics=metrics.classification_report(Y_train,LR_Train_predict,output_dict=True)
Df_New=pd.DataFrame(LR_train_metrics).T
Df_New
```
