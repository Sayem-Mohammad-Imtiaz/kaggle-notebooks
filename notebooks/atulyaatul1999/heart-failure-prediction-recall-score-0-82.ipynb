```python
y=df['DEATH_EVENT']
```

```python
y
```

```python
X=df.drop(columns=['DEATH_EVENT'])
```

```python
X
```

```python
X=X.values
```

```python
# One row of data after Standardization
X[0]
```

### we need to shuffle the data as some algorithms doesn't work well when we passes data of one category in one batch and onther category data in another batch

```python
from sklearn.utils import shuffle
```

```python
X,y= shuffle(X,y, random_state=42)
```

### Spliting the data in Train and test data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Training a logistic regression model

```python
from sklearn.linear_model import LogisticRegression
```

```python
le=LogisticRegression()
```

```python
le.fit(X_train,y_train)
```

### best fitting intercept and coefficients

```python
print(le.intercept_,le.coef_)
```

```python
y_pred=le.predict(X_test)
```

```python
from sklearn.metrics import recall_score,precision_score,confusion_matrix,f1_score
```

```python
cm=confusion_matrix(y_test,y_pred)
recall=recall_score(y_test,y_pred)
precision=precision_score(y_test,y_pred)
```

```python
cm
```

```python
# Confusion matrix 
sns.heatmap(cm, annot=True)
```

```python
print(recall,precision)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

### We get accuracy of 8686% is it good, No as we have recall score of only 0.72 but we want to maximize the recall score because we want to minimize false negative predictions (We want our model to predict each Heart Failure event correctly)

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(max_depth=6)
```

```python
dt.fit(X_train,y_train)
```

```python
y_pred2=dt.predict(X_test)
```

```python
recall_score(y_test,y_pred2)
```

```python
precision_score(y_test,y_pred2)
```

```python
f1_score(y_test,y_pred2)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
confusion_matrix(y_test,y_pred2)
```

```python
sns.heatmap(confusion_matrix(y_test,y_pred2),annot=True)
```

### in this descision tree model we are getting accuracy score of 0.777 but its recall score is 0.827, recall score should be high in this case so we will choose descision tree model over logistic regression model.

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=800)
```

```python
recall_score(y_test,rf.predict(X_test))
```
