## Indexing, Selecting and Slicing Dataframe 

In this section lets see how to get back data fram a selected columns or even multiple columns

First lets see names all games ranked in this dataframe 

```python
gmsales_df.Name
```

```python
gmsales_df['Name']
```

Both methods shown above do the same this but the secon method is preferable if a column name contains spaces.

Each column is represented using a data structure called Series, which is essentially a numpy array with some extra methods and properties.

```python
type(gmsales_df['Name'])
```

Now to get multiple columns

```python
gm_release = ['Name', 'Year']
```

```python
gmsales_df[gm_release]
```

> Or we can just use this method:

```python
gmsales_df[['Name', 'Year']]
```

We have seen how to get columns, now lets see how to get rows as well

**loc[]** can be used to get rows by index name

```python
gmsales_df.loc[4]
```

in this dataframe index names are numbers same as an array. But if we define names to these rows we can put the names in loc[] in ''.
But since we dont have name in indexes we can just use iloc[] method in which we have to put index number. We use it if we have to get 10th or 77th row of a dataframe. This is also store in numpy array format starting from 0 to n - 1.

```python
gmsales_df.iloc[4]
```

```python
gmsales_df.iloc[-1]
```

Just like in python arrays [-1] will give you last row

```python
gmsales_df.loc[[4, 7]]
```

```python
gmsales_df.iloc[[4, 7]]
```

Okay great! Now lets see how get data based on specific row and column. For this we can just use

```python
gmsales_df['Name'][7]
```

or we can also use at() function

```python
gmsales_df.at[7,'Name']
```
