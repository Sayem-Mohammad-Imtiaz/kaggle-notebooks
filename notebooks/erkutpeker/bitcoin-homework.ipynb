```python
data = pd.read_csv('../input/bitcoin-historical-data/bitstampUSD_1-min_data_2012-01-01_to_2020-09-14.csv')
data.head()
```

```python
data.corr()
```

```python
# correlation map
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
# correlation = ilişki oran orantı gibi
# annot = True demek kutucukların içindeki sayıların gözükmesi
```

```python
# line plotda gorsellestirelim
data.Open.plot(kind = "line" , color = "r" , label = "Open",linewidth=1,alpha = 0.7,grid = True,linestyle=":")
data.Close.plot(color = "b" , label = "Close",linewidth=1,alpha = 0.7,grid = True,linestyle=":")
plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('x axis')              # label = name of label
plt.ylabel('y axis')
plt.title('bitcoinin fiyati')            # title = title of plot
plt.show()



  
```

SCATTER

```python
# scatter plotda gorsellestirelim
data.plot(kind="scatter",x = "High",y = "Low",alpha = 0.4,color="black")
plt.xlabel("High")
plt.ylabel("Low")
plt.title("btc fiyati")

```
