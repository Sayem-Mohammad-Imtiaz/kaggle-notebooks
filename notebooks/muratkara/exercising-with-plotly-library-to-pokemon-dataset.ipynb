## Now starting first comparing about First Generation Pokemons to Attack's and Defense

```python
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)


trace1 = go.Scatter( x = generation1.Row,
                     y = generation1.Attack,
                     mode = "lines+markers",
                     name = "Attacks",
                     marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                     text = generation1.Name)
trace2 = go.Scatter( x = generation1.Row,
                     y = generation1.Defense,
                     mode = "lines+markers",
                     name = "Defense",
                     marker = dict(color = 'rgba(80, 26, 80, 0.2)'),
                     text = generation1.Name)
data = [trace1,trace2]
layout = dict(title = 'Attack vs Defense of First Generation Pokemons', xaxis = dict(title = 'First Generation Pokemons',ticklen=5,zeroline=False))
fig = dict(data = data, layout = layout)
iplot(fig)
```

## Now i want show you number of First,Second,Third,Fourth.. generation pokemons numbers(Using Piechart)

```python
number_of_generation1 = df[df["Generation"]==1].count(axis=1)
number_of_generation2 = df[df["Generation"]==2].count(axis=1)
number_of_generation3 = df[df["Generation"]==3].count(axis=1)
number_of_generation4 = df[df["Generation"]==4].count(axis=1)
number_of_generation5 = df[df["Generation"]==5].count(axis=1)
number_of_generation6 = df[df["Generation"]==6].count(axis=1)
```

```python
number_of_generation1 = len(number_of_generation1)
number_of_generation2 = len(number_of_generation2)
number_of_generation3 = len(number_of_generation3)
number_of_generation4 = len(number_of_generation4)
number_of_generation5 = len(number_of_generation5)
number_of_generation6 = len(number_of_generation6)
```

```python
new_df = pd.DataFrame(np.array([[number_of_generation1,number_of_generation2,number_of_generation3,number_of_generation4,number_of_generation5,number_of_generation6]]))
```

```python
new_df
```

```python
new_df = new_df.T
```

```python
new_df
```

```python
new_df.rename(index={0:'First Generation',1:'Second Generation',2:'Third Generation',3:'Fourth Generation',4:'Fifth Generation',5:'Sixth Generation'},inplace=True)
```

```python
new_df
```

```python
new_df.rename(columns={0:'Numbers'},inplace=True)
```

```python
new_df
```

```python
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)

pie1 = (new_df.Numbers)
labels = [x for x in new_df.index.values]
fig = {
     "data":[
         {
             "values":pie1,
             "labels":labels,
             "domain":{"x":[0, .5]},
             "name":"Number of Pokemons Generations",
             "hoverinfo":"label+percent+name",
             "hole": .3,
             "type": "pie"
         },
         
     ],
    "layout":{
        "title":"Pokemons Generation Numbers",
        "annotations": [
            {
                "font":{"size":20},
                "showarrow":False,
                "text": "Number of Pokemons",
                "x": 0.20,
                "y": 1
            },
        ]
    }
}
iplot(fig)
```

```python
trace1 = go.Scatter3d( x = legendary_generation3.Attack,
                       y = legendary_generation3.Speed,
                       z = legendary_generation3.Defense,
                       mode = 'markers',
                       marker=dict(
                             size = 10,
                             color = 'rgb(255,0,0)',
                       )
                     )
data = [trace1]
layout = go.Layout( margin=dict( l=0,
                                 r=0,
                                 b=0,
                                 t=0
                               )
                  )
fig = go.Figure(data=data,layout=layout)
iplot(fig)
```

## And Comparing first three generation pokemons attack's(using marker plot)

```python
import plotly.graph_objs as go

trace1 = go.Scatter( x = generation1.Row,
                     y = generation1.Attack,
                     mode = "markers",
                     name = "First Generation",
                     marker = dict(color = 'rgba(255, 128, 255, 0.8)'),
                     text = generation1.Name)
trace2 = go.Scatter( x = generation2.Row,
                     y = generation2.Attack,
                     mode = "markers",
                     name = "Second Generation",
                     marker = dict(color = 'rgba(255, 128, 2, 0.8)'),
                     text = generation2.Name)
trace3 = go.Scatter ( x = generation3.Row,
                      y = generation3.Attack,
                      mode = "markers",
                      name = "Third Generation",
                      marker = dict(color = 'rgba(0, 255, 200, 0.8)'),
                      text = generation3.Name)
data = [trace1,trace2,trace3]
layout = dict(title = 'Top 3 Pokemon Generations Comparing Attacks',
              xaxis = dict(title = 'Rows',ticklen=5,zeroline=False),
              yaxis = dict(title = 'Attacks',ticklen=5,zeroline=False)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```
