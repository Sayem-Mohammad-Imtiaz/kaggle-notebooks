```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('../input/sms-spam-collection-dataset/spam.csv', encoding = 'latin-1')
df.head()
```

```python
df.shape
```

```python
df.groupby('v1').describe()
```

```python
df1=df[['v1','v2']]
df1.columns=['spam','email']
df1.head()
```

```python
df1['spam'].apply(lambda x:1 if x=='spam' else 0)
df1.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df1.email,df1.spam,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
#vectorizer = CountVectorizer()

from sklearn.naive_bayes import MultinomialNB
#mb=MultinomialNB()
```

```python
from sklearn.pipeline import Pipeline
clf= Pipeline([
    ('vectorizer',CountVectorizer()),
    ('mb',MultinomialNB())
    
])
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
y_predict= clf.predict(X_test)
y_predict
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,y_predict)
cm
```

```python
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True,cmap="YlGnBu",fmt=".1f")
plt.xlabel("Prediction")
plt.ylabel("Truth")
```

```python
from sklearn.metrics import classification_report
classification_report(y_test,y_predict,output_dict=False)
```

```python
print(classification_report(y_test, y_predict))
```
