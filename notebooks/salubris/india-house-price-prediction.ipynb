```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
india_data = pd.read_csv("../input/house-price-prediction-challenge/train.csv")
```

```python
india_data.head()
```

```python
# Choosing a target variable stored in a new variable. 
y = india_data["TARGET(PRICE_IN_LACS)"]

# Choosing which features to be included
india_features = ["UNDER_CONSTRUCTION", "RERA", "BHK_NO.", "SQUARE_FT", "READY_TO_MOVE", "RESALE"]
X = india_data[india_features]
```

```python
# Split into validation and training data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
# Specify model
india_model = DecisionTreeRegressor(random_state=1)
india_model.fit(train_X, train_y)
```

```python
# Validate
val_preds = india_model.predict(val_X)
val_mae = mean_absolute_error(val_preds, val_y)
print("Validation MAE when not specifying max_leaf_nodes: {:,.0f}".format(val_mae))
```

```python
# Best value for max_leaf_nodes
india_model = DecisionTreeRegressor(max_leaf_nodes=150, random_state=1)
india_model.fit(train_X, train_y)
val_preds = india_model.predict(val_X)
val_mae = mean_absolute_error(val_preds, val_y)
print("Validation MAE for best value of max_leaf_nodes: {:,.0f}".format(val_mae))
```

```python
rf_india = RandomForestRegressor(random_state=1)
rf_india.fit(train_X, train_y)
rf_val_preds = rf_india.predict(val_X)
rf_val_mae = mean_absolute_error(rf_val_preds, val_y)
print("Validation MAE for Random Forest Model: {:,.0f}".format(rf_val_mae))
```

```python
# To improve accuracy, create a new Random Forest model which you will train on all training data
rf_model_on_full_data = RandomForestRegressor()

# fit rf_model_on_full_data on all data from the training data
rf_model_on_full_data.fit(X, y)
```

```python
# Implement test data
test_data = pd.read_csv("../input/house-price-prediction-challenge/test.csv")
```

```python
test_X = test_data[india_features]
```

```python
test_preds = rf_model_on_full_data.predict(test_X)
```
