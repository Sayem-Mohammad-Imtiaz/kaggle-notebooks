# Introduction

In this notebook, I will be showing you how to build a SMS Spam Classifier, which will be able to predict whether an SMS (English) is **ham** or **spam**. I will be using the UCI Machine Leaning Repository's [SMS Spam Colletion Dataset](https://www.kaggle.com/uciml/sms-spam-collection-dataset). So, lets have a look on how to build a SMS spam classifier using NLP.

![Spam](https://zpllp.com/wp-content/uploads/2018/04/no-spam-300x296.jpg)

# Importing Libraries

```python
import pandas as pd
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

# Importing Dataset

```python
df = pd.read_csv('../input/sms-spam-collection-dataset/spam.csv',encoding='ISO-8859-1')
df.head()
```

# Dropping Unnecessary Columns

```python
df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'],axis='columns',inplace=True)
df.columns = ['label', 'message']
df.head()
```

# Tokenization & Stemming

```python
ps = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(df)):
    review = re.sub('[^a-zA-Z]', ' ', df['message'][i])
    review = review.lower()
    review = nltk.word_tokenize(review)
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
    
    
```

# Implement Bag Of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000)
X = cv.fit_transform(corpus).toarray()
```

```python
y = pd.get_dummies(df['label'])
y = y.iloc[:,1].values
```

```python
print(df['label'].head(),"\n y = ",y)
```

This shows that **0 indicates ham** and **1 indicates spam**

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2, random_state=0)
```

# Naive Bayes Model

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model = MultinomialNB()
spam_detect_model.fit(X_train,y_train)
```

```python
y_pred = spam_detect_model.predict(X_test)
```

# Accuracy & Result

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
acc_score = accuracy_score(y_test,y_pred)
acc_score
```

# Conclusion

The accuracy is pretty good. So, this model can be a good SMS Spam Classifier.
