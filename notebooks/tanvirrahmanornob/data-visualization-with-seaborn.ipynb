```python
import numpy             as  np
import pandas            as pd
import seaborn           as sns
import matplotlib.pyplot as plt
```

```python
athlete_data = pd.read_csv("../input/olympic-games/athletes.csv")
country_data = pd.read_csv("../input/olympic-games/countries.csv")
tips_data    = sns.load_dataset("tips")
```

```python
## the relation between the population and GDP
sns.relplot(x='population',y='gdp_per_capita',data=country_data)
```

```python
athlete_data.head(3)
```

```python
### what is the relation between height and weight among the athlete
sns.relplot(x="height",y="weight",data = athlete_data)
```

```python
## relation of height and weight based on sex
sns.relplot(x="height",y="weight",hue="sex",data=athlete_data)
```

```python
## relation of height and weight based on sex
sns.relplot(x="height",y="weight",hue="sport",height=7,aspect=2,data=athlete_data)
## what is the relation between x and y based on hue
```

```python
## categorial data
## relation between the day and the tip
sns.catplot(x="day",y="tip",data=tips_data)
```

```python
sns.catplot(x="day",y="tip",kind="swarm",data=tips_data)
```

```python
## relation between tips and day based on sex
sns.catplot(x="day",y="tip",hue="sex",data=tips_data)
```

```python
## relation between tips and day based on sex
sns.catplot(x="day",y="tip",kind="box",data=tips_data)
```

```python
sns.catplot(x="day",y="tip",hue="sex",kind="box",data=tips_data)
```

```python
sns.catplot(x="day",y="tip",hue="sex",kind="violin",data=tips_data)
```

```python
# what is the distribution of the total bil
sns.distplot(tips_data['total_bill'])
```

```python
## most of the tips are from 15 to 20,21
## now what are the distribution of the tips
sns.distplot(tips_data['tip'])
```

```python
# 2 to 3 are the most of the tips
## the most important 
## what is the relation between the tips and the total bill
## so we need to join the two plot
sns.jointplot(x=tips_data['total_bill'],y=tips_data['tip'])
```
