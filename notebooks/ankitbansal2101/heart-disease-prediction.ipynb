```python
#reading the data
data=pd.read_csv("/kaggle/input/heart-disease-uci/heart.csv")
data.head()
```

```python
#age,sex,cp,trestbps,chol,fbs,restecg,thalach,exang,oldpeak,slope,ca,thai columns are featurd array
#target column is our target variable
```

```python
data.describe()
```

```python
# checking null values
data.isnull().sum()
#no null values present
```

```python
#creating feature matrix and target array
y=data["target"]
X=data.drop("target",axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.3)
```

```python
#normalising the data
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
scaler.fit(X_train)
X_train_scaled=scaler.transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

**logistic regression**
****

```python
#make model
model=LogisticRegression()
model.fit(X_train_scaled,y_train)
```

```python
#making predictions
y_pred=model.predict(X_test_scaled)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
#accuracy using confusion matrix
accuracy=(33+40)/(33+40+8+10)
accuracy
```

# knn classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=8)
knn.fit(X_train_scaled,y_train)

```

```python
y_predict_knn=knn.predict(X_test_scaled)
y_predict_knn
```

```python
confusion_matrix(y_predict_knn,y_test)
```

```python
acc_knn=69/91
acc_knn
```

```python
knn.score(X_test_scaled,y_test)
```

# using different value of n_neighbors to find optimal value of n_neighbors

```python
accuracy_knn=[]
for k in range(1,15):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled,y_train)
    accuracy_knn.append([k,knn.score(X_test_scaled,y_test)])
    print(k,knn.score(X_test_scaled,y_test))

    
```

# we can see that for k=9 , our accuracy is max i.e 81.3%

```python
accuracy_knn
```
