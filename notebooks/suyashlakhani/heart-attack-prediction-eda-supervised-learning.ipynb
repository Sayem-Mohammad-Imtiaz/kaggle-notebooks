# Model Building and Standardization

```python
X = df1.iloc [ : , : -1].values
Y = df1.iloc [ :, -1].values
X
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
x_train = ss.fit_transform(x_train)
x_test = ss.transform(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from xgboost import XGBRegressor
from catboost import CatBoostClassifier
```

```python
acc_score = []
model = [LogisticRegression(),SVC(),KNeighborsClassifier(),DecisionTreeClassifier(),RandomForestClassifier(),GaussianNB()]
for m in model:
    m.fit(x_train,y_train)
    pred = m.predict(x_test)
    score = accuracy_score(y_test,pred)
    acc_score.append(score)
```

```python
acc_score
```

```python
model_names = ['LogisticRegression', 'SVC','KNN', 'Decisiontree','RandomForest','GaussianNB']
```

```python
sns.barplot(x=acc_score, y=model_names, palette='mako')
```

##  **LogisticRegression performed the best with an accuracy of 81.3% **
