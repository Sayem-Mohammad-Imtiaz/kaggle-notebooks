```python
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier 
from sklearn.ensemble import ExtraTreesClassifier 
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
model_name = [DecisionTreeClassifier,RandomForestClassifier,ExtraTreesClassifier,XGBClassifier,LogisticRegression]
from sklearn.metrics import accuracy_score
```

```python
df1.head()
```

```python
x = df1.drop(["quality"], axis =1)
y = df1['quality']
```

```python
x.head()
```

```python
x.shape
```

Handling Imbalanced Dataset

```python
#scaler  =StandardScaler()
#x = scaler.fit_transform(x)
from imblearn.over_sampling import SMOTE
oversample = SMOTE(k_neighbors=4)
# transform the dataset
x, y = oversample.fit_resample(x, y)
x.shape
```

```python
from sklearn.model_selection import cross_val_score, train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

# Model Prediction XGBOOST



```python
model = XGBClassifier()
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
score = accuracy_score(y_test,pred)
score
```

# Model using ANN

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = keras.Sequential([
    layers.BatchNormalization(input_shape=[11]),
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(512, activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(1),
])
```

```python
model.compile(
optimizer ="adam",
loss = "mae"
    
)
```

```python
history = model.fit(x_train,y_train, validation_data = (x_test,y_test), batch_size = 512, epochs = 100)
```

```python
history_df = pd.DataFrame(history.history)
history_df
```

```python
history_df.loc[:100, "loss"].plot()
history_df.loc[:100, "val_loss"].plot()
```
