# Data Cleaning and preprocessing 
1. Removing stop words(using stopwords) and ,!? etc using Regex
2. Converting to lower case and spliting the messages into words
3. Applying Lematization/stemming to words (excluding stop words)
4. Join the words back into a sentence
5. Store in a new list

```python
import nltk 
import re
from nltk.corpus import stopwords #Stop words finder
from nltk.stem.porter import PorterStemmer #Stemming the words

```

```python
ps = PorterStemmer()
corpus = []   # Saving the final sentences (step5)
```

```python
for i in range(0,len(messages)):
    review = re.sub('[^a-zA-Z]'," ", messages["v2"][i])  #Removing special characters etc
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if word not in stopwords.words('english')]  #Stemming words which are not stopwords
    review = " ".join(review)
    corpus.append(review)
    
```

# Converting words to vectors (Bag of words) for model to understand

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500)
x = cv.fit_transform(corpus).toarray()
```

```python
y = pd.get_dummies(messages["v1"],drop_first=True)
y
```

```python
y = np.array(y)
```

```python
y
```

# Model Building

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(x_train, y_train)
```

```python
y_pred=spam_detect_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_pred,y_test)
```

```python
score
```

```python
from sklearn.model_selection import cross_val_score
cv_score = cross_val_score(spam_detect_model,x,y,cv=5)
print("Cross Val scoe: {}".format(np.mean(cv_score)))
```
