```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 

```

```python
df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv')
df.head()
```

```python
df.info()
df.shape
```

```python
#looking for NULL values
df.isnull().sum(axis=0)
```

```python
#dropping souce_name & source_website
df.drop(['source_name', 'source_website'], axis=1, inplace=True)
df.head()
```

```python
df.dtypes
```

```python
#converting date to date
df['date'] = pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
#total vaccinations
df['total_vacc'] = df.groupby('country')['total_vaccinations'].tail(1)

#top 10 country with total vaccinations
x = df.groupby('country')['total_vacc'].mean().sort_values(ascending=False).head(10)

#bargraph
plt.figure(figsize=(7,7))
sns.set_theme(style="whitegrid")
sns.barplot(x=x.values, y=x.index, orient='h',palette='Blues_r')
plt.xlabel("Total Vaccinations")
plt.ylabel("Country")
print(x)

```

```python
#total people fully vaccinated
df['total_fully_vacc'] = df.groupby('country')['people_fully_vaccinated'].tail(1)

#top 10 country with total people fully vaccinated
x = df.groupby('country')['total_fully_vacc'].mean().sort_values(ascending=False).head(10)

#bargraph
plt.figure(figsize=(7,7))
sns.set_theme(style="whitegrid")
sns.barplot(x=x.values, y=x.index, orient='h',palette='Blues_r')
plt.xlabel("Total Fully Vaccinated")
plt.ylabel("Country")
print(x)
```

```python
#total people fully vaccinated per hundred
df['total_fully_vacc_per_100'] = df.groupby('country')['people_fully_vaccinated_per_hundred'].tail(1)

#top 10 country with total people fully vaccinated per hundred
x = df.groupby('country')['total_fully_vacc_per_100'].mean().sort_values(ascending=False).head(10)

#bargraph
plt.figure(figsize=(7,7))
sns.set_theme(style="whitegrid")
sns.barplot(x=x.values, y=x.index, orient='h',palette='Blues_r')
plt.xlabel("Total Fully Vaccinated Per Hundred")
plt.ylabel("Country")
print(x)
```

```python
#total vaccinations
plt.figure(figsize=(10,10))
sns.lineplot(x='date', y='total_vaccinations', data=df)
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
```

```python
#daily vaccinations trend
x = df.groupby('date')['daily_vaccinations'].sum()

plt.figure(figsize=(10,10))
sns.lineplot(x=x.index, y=x.values)
plt.title('Daily Vaccinations Trend')
plt.xlabel("Date")
plt.show()
```

```python
#most common vaccines
df['vaccines'].value_counts()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(y='vaccines', data=df, palette='Blues_r')
plt.ylabel('Vaccines')
plt.xlabel('Count')
plt.title('Most Common Vaccine Combinations Used')
plt.show()
```
