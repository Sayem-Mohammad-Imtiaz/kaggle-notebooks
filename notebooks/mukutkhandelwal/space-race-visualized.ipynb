## Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style(style = 'whitegrid')
```

## Reading Data

```python
df = pd.read_csv('../input/all-space-missions-from-1957/Space_Corrected.csv')
```

## First 5 rows of the data

```python
df.head()
```

## 2. Preprocessing

### 2.1. Checking the null values through heatmap

```python
sns.heatmap(df.isnull())
```

### 2.2. Summary of null values in columns

```python
df.isnull().sum()
```

### 2.3. Droping the columns which are not required

```python
df.drop(columns=[' Rocket','Unnamed: 0','Unnamed: 0.1'],axis = 'columns',inplace=True)
```

### 2.4 Extracting the Country name and Year from location and datum column

### 2.4.1. Country

```python
def county(x):
  a = x.split(',')
  return a[-1]
df['Country'] = df['Location'].apply(county)
```

### 2.4.2. Year

```python
def years(x):
  x = pd.to_datetime(x)
  return x.year
df['Year']= df['Datum'].apply(years)
```

## Data after Preprocessing

```python
df.head(3)
```

## 3. Data Visualization ðŸ“‰

### 3.1. Q1 How many missions have carried out by organizations since 1957?
### Q2 Which organizations have maximum misssions since 1957 ?

```python
total_company = df.groupby('Company Name')['Country']
comp = []
comp_count = []
for i in total_company:
  comp.append(i[0])
  comp_count.append(len(i[1]))
df_comp_count = pd.DataFrame({'Company':comp,'Total Company': comp_count})
```

### On an average of 77.21 missions has been carried out by the organizations and with RVSN USSR being carrying out maximun number of missions

```python
df_comp_count[:10]
```

```python
plt.figure(figsize = (12,15))
df_comp_count = df_comp_count.sort_values('Total Company',ascending =False)
sns.barplot(x= 'Total Company',y = 'Company',data = df_comp_count )
plt.title('Total number of Mission Organization wise since 1957')
```

## 3.2. Which countries have most number of successful space missions
### Top 3 being:-
### 1.  Russia
### 2.  USA
### 3.  Kazakhstan

```python
df_1 = df[df['Status Mission'] == 'Success']
plt.figure(figsize=(12,15))
sns.countplot(x = 'Country',hue= 'Status Mission',data= df_1,order = df_1['Country'].value_counts().index)
plt.xticks(rotation= 90)
plt.title('Total Successful Mission Country wise')
```

## 3.3. Which countries have most number of Failure space missions
### Top 3 being:-
### 1. USA
### 2. Kazakhstan
### 3. Russia

```python
df_2 = df[df['Status Mission'] == 'Failure']
plt.figure(figsize=(12,15))
sns.countplot(y = 'Country',hue= 'Status Mission',data= df_2,order=df_2['Country'].value_counts().index)
plt.title('Total Mission Failed')
plt.xticks(rotation= 90)
```

### 3.4 Which year has the maximum number of space missions since 1957?
### top 3 being
### 1. 1971
### 2. 2018
### 3. 1977

```python
plt.figure(figsize=(12,18))
sorted_year = df.sort_values('Year',ascending = False)
sns.countplot(y = 'Year',data = sorted_year,order = df['Year'].value_counts().index)
plt.title('Total number of Space Missions Year wise')
```

## 3.5 What are the Status of the rocket witch respect to the organization handling them?

### We can see that RVSN USSR has retired all of its rocket the main reason being collapse of USSR

```python
plt.figure(figsize = (8,18))
sns.countplot(y = 'Company Name',hue = 'Status Rocket',data = df,order = df['Company Name'].value_counts().index)
plt.legend()
plt.title('Status of rocket according to company ')
```
