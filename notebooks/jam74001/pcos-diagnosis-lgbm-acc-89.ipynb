Dropping the repetitive features:

```python
data.drop(['Unnamed: 44', 'Sl. No_y', 'PCOS (Y/N)_y', '  I   beta-HCG(mIU/mL)_y',
       'II    beta-HCG(mIU/mL)_y', 'AMH(ng/mL)_y'], axis=1, inplace= True)
```

```python
data["AMH(ng/mL)"]= pd.to_numeric(data["AMH(ng/mL)"], errors='coerce')
data['II    beta-HCG(mIU/mL)']= pd.to_numeric(data['II    beta-HCG(mIU/mL)'], errors='coerce')
```

```python
data.isnull().sum()
```

Dealing with non numeric values in the data:

Dealing with missing values:


```python
data['Marraige Status (Yrs)']= data['Marraige Status (Yrs)'].fillna(data['Marraige Status (Yrs)'].mode()[0])
data['Fast food (Y/N)']= data['Fast food (Y/N)'].fillna(data['Fast food (Y/N)'].mode()[0])
data['AMH(ng/mL)']= data['AMH(ng/mL)'].fillna(data['AMH(ng/mL)'].mode()[0])
data['II    beta-HCG(mIU/mL)']= data['II    beta-HCG(mIU/mL)'].fillna(data['II    beta-HCG(mIU/mL)'].mode()[0])
```

```python
y= data['PCOS (Y/N)']
X= data.drop('PCOS (Y/N)', axis=1)
import re
X = X.rename(columns = lambda x:re.sub('[^A-Za-z0-9_]+', '', x))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=42)
X_smote, y_smote = sm.fit_resample(X_train, y_train)
```

```python

model = LGBMClassifier(max_bin= 500, learning_rate=0.1, num_leaves=500)
model.fit(X_smote, y_smote)
# cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=2, random_state=1)
# n_scores = cross_val_score(model2, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
# print('Accuracy: %.3f (%.3f)' % (np.mean(n_scores), np.std(n_scores)))
```
