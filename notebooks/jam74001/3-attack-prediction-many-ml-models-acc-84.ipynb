```python
X= data.drop('output', axis=1)
y= data['output']
```

```python
model = xgb.XGBClassifier()
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=2, random_state=1)
n_scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
print('Accuracy: %.3f (%.3f)' % (np.mean(n_scores), np.std(n_scores)))
```

```python
model2 = LGBMClassifier()
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=2, random_state=1)
n_scores = cross_val_score(model2, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
print('Accuracy: %.3f (%.3f)' % (np.mean(n_scores), np.std(n_scores)))
```

```python
from sklearn.linear_model import LogisticRegression
model3= LogisticRegression()
n_scores = cross_val_score(model3, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
print('Accuracy: %.3f (%.3f)' % (np.mean(n_scores), np.std(n_scores)))
```

```python
from sklearn.ensemble import RandomForestClassifier
model4= RandomForestClassifier()
n_scores = cross_val_score(model4, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
print('Accuracy: %.3f (%.3f)' % (np.mean(n_scores), np.std(n_scores)))
```

```python
from sklearn.tree import DecisionTreeClassifier
model5= DecisionTreeClassifier()
n_scores = cross_val_score(model5, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
print('Accuracy: %.3f (%.3f)' % (np.mean(n_scores), np.std(n_scores)))
```
