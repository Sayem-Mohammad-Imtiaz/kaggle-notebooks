**Task: Which US state has the highest number of accidents,and a description of the accidents that usually occur in that state.**

### Load Data

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

df = pd.read_csv("/kaggle/input/us-accidents/US_Accidents_May19.csv")
```

### Explore few rows to get a sense of data

```python
df.shape
```

### There are 2.24 Million records in data . Let's check if there are any blanks.

```python
df.isnull().sum()
```

### Ok , there are missing values but not in state , id columns . So , we are good to go

### Approach : We will do a group by on state and do a sum to get total accidents per state . We will also create a bar chart to show our findings and write a summary note . Let's begin the fun journey.

```python
state_wise_counts = df.groupby('State')['ID'].count().reset_index()
```

```python
state_wise_counts.shape
```

### There are 49 rows for 49 states.

```python
state_wise_counts.head()
```

### Counts are not sorted . Let's sort in descending order.

```python
state_wise_counts = state_wise_counts.sort_values(by = "ID",ascending=False)
```

```python
state_wise_counts.head()
```

### Cool , we can see that California has max accidents . Lets plot it through searborn library and see the results.

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
```

```python
f, ax = plt.subplots(figsize=(6, 15))
sns.barplot(y="State", x="ID", data=state_wise_counts)
```

### We can see that California tops the list .Texas , Florida , North Carolina , New york follows it . Major accident happens in big states . It would be interesting to find accident rate per 1000 people and see which has highest rate . I am sure insurance premium would be higher in those states.
