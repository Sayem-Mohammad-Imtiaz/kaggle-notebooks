```python
#reads the dataset
star_data = pd.read_csv("/kaggle/input/star-type-classification/Stars.csv")
#allowed_types = [1,2]
#star_data = star_data[star_data['Type'].isin(allowed_types)]
print(star_data)
from sklearn.model_selection import train_test_split
train_data, test_data = train_test_split(star_data, test_size=0.25)
```

```python
#defines the targets and variables 
features = np.array(train_data[['A_M', 'R']])
targets = np.array(train_data['Type'])
                    
test_features = np.array(test_data[['A_M', 'R']])
test_targets = np.array(test_data['Type'])
```

```python
#makes a knn model
from sklearn.neighbors import KNeighborsClassifier
#the model will look at the 5 nearest points, wieghted by distance
knn_model = KNeighborsClassifier(5, weights='distance')
knn_model.fit(features, targets)
```

```python
import matplotlib.pyplot as plt
labels, index = np.unique(targets, return_inverse=True)

#plots the data it's predicting with absolute magnitude on the x axis and radius on the y 
scatter = plt.scatter(features[:,0], features[:,1], c=index)
plt.legend(scatter.legend_elements()[0], labels)
plt.show
```

```python

predictions = knn_model.predict(test_features)

correct = 0
incorrect = 0
#the loop will predict each one and reocrd if it was right or wrong, and calculte the percentage at the end
for i in range(len(predictions)):
    print(f"Predicted {predictions[i]}, actually {test_targets[i]}.")
    if predictions[i] == test_targets[i]:
        correct +=1
    else:
        incorrect += 1
        print('Wrong')
print(f"Got {correct} correct and {incorrect} incorrect.")
print(f"{100*correct/(correct+incorrect)} percent accuracy")
```
