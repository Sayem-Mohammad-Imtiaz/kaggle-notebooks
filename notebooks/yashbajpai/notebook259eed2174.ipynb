```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

# DATA PREPROCESSING

```python
df_heart = pd.read_csv('../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv')
df_heart.head()
```

```python
X = np.asarray(df_heart.drop(columns = ['DEATH_EVENT']))
Y = np.asarray(df_heart['DEATH_EVENT'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3)
```

# ** Scaling Data **

```python
slr = StandardScaler()
slr.fit(x_train)
x_train = slr.transform(x_train)
x_test = slr.transform(x_test)
```

```python
print(x_train.shape, y_train.shape)
```

```python
print(x_test.shape, y_test.shape)
```

# Linear - Support Vector Machine Model

```python
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
svm_model = svm.SVC(kernel = 'linear')
svm_model.fit(x_train, y_train)
```

```python
svm_model_prediction = svm_model.predict(x_train)
```

```python
print('accuracy of training data predictions :' , accuracy_score(y_train, svm_model_prediction))
```

```python
svm_model_testPrediction = svm_model.predict(x_test)
```

```python
print('accuracy of testing data predictions :', accuracy_score(y_test, svm_model_testPrediction))
```

# Logistic Regression Model


```python
from sklearn.linear_model import LogisticRegression
Logistic_Model = LogisticRegression(C = 0.01, solver = 'liblinear').fit(x_train, y_train)
Logistic_Model_Prediction = Logistic_Model.predict(x_train)
```

```python
print("Accuracy of training data prediction :", accuracy_score(y_train, Logistic_Model_Prediction))
```

```python
Logistic_Model_TestPrediction = Logistic_Model.predict(x_test)
```

```python
print('Accuracy of test data prediction :', accuracy_score(y_test, Logistic_Model_TestPrediction))
```

# Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
DT_Model = DecisionTreeClassifier(criterion = 'entropy', max_depth = 4)
DT_Model.fit(x_train, y_train)
DT_Model_Prediction = DT_Model.predict(x_train)
```

```python
print('Acccuracy of training data prediction : ', accuracy_score(y_train, DT_Model_Prediction))
```

```python
DT_Model_TestPrediction = DT_Model.predict(x_test)
```

```python
print("Accuracy of test data predictrion :", accuracy_score(y_test, DT_Model_TestPrediction))
```
