```python
import seaborn as sns
```

```python
highest = [ind.Runs.max(),aus.Runs.max(),sa.Runs.max(),eng.Runs.max(),sl.Runs.max(),wi.Runs.max(),pak.Runs.max(),nz.Runs.max(),zim.Runs.max(),bdesh.Runs.max(),afg.Runs.max(),ire.Runs.max()]
```

```python
average = [ind.Runs.mean(),aus.Runs.mean(),sa.Runs.mean(),eng.Runs.mean(),sl.Runs.mean(),wi.Runs.mean(),pak.Runs.mean(),nz.Runs.mean(),zim.Runs.mean(),bdesh.Runs.mean(),afg.Runs.mean(),ire.Runs.mean()]
```

```python
lowest = [ind.Runs.min(),aus.Runs.min(),sa.Runs.min(),eng.Runs.min(),sl.Runs.min(),wi.Runs.min(),pak.Runs.min(),nz.Runs.min(),zim.Runs.min(),bdesh.Runs.min(),afg.Runs.min(),ire.Runs.min()]
```

```python
scores = pd.DataFrame(list(zip(countriesbarmore, highest, average, lowest)), columns =['Countries', 'highest', 'average','lowest'])
```

```python
scores
```

```python
fig, axs = plt.subplots(ncols=3)
fig.set_size_inches(25, 5)
sns.barplot(x='Countries', y='highest', data=scores, ax=axs[0])
sns.barplot(x='Countries', y='average', data=scores, ax=axs[1])
sns.barplot(x='Countries', y='lowest', data=scores, ax=axs[2])
plt.show() 
```

These graphs compare the highest, average and the lowest runs by a player from each country

```python
plt.plot( 'Countries', 'highest', data=scores, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 'Countries', 'average', data=scores, marker='', color='red', linewidth=2, linestyle='dashed',)
plt.plot( 'Countries', 'lowest', data=scores, marker='', color='green', linewidth=2,  label="lowest")
plt.legend()
```

This graph depicts highest lowest and average on one graph
