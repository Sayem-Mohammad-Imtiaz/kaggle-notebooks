```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split



import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
df=pd.read_csv('../input/star-type-classification/Stars.csv')

df.shape
```

```python
df.head()
```

It's always a good idea to get some info on your dataset before making any predictions. I'll start by checking for missing values and how well the dataset is balanced.

```python
df.isna().sum()
```

Great! no missing values. 

```python
df.groupby('Type').count().plot.bar()
```

It appears that the data is well balanced. The next step in this process is to preprocess the data. 

```python
#setting X and Y variables.

X = df[['Temperature', 'L', 'R', 'A_M']]
y = df['Type']
```

```python
#Setting up train and test data.

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
#applying scaling to data

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train) 
X_test = scaler.fit_transform(X_test)
```

# Applying Kmeans clustering

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()

model.fit(X_train, y_train)
```

```python
K_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, K_pred))


```

The performance for most classifications is quite good. With some exceptions for 0 and 1.
