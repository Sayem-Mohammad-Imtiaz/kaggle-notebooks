# TV Shows and Movies Distribution

```python
# TV shows and movies distribution
import matplotlib.pyplot as plt

shows_count=df["type"].value_counts().sort_index()
#shows_count
#add colors
colors = ['#66b3ff','#ffcc99']
plt.figure(figsize = (10,6))
plt.pie(df['type'].value_counts(), labels = df['type'].value_counts().index,colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Movies and TV Shows Distribution")
plt.show()
```

# Overall Rating count(Shows & Movies)

```python
# Overall Rating count both TV Shows and Movies

ratings = df["rating"].value_counts().sort_index()
ratings
```

```python
# Distribution showing overall ratings
import seaborn as sns

plt.figure(figsize = (12,12))
bars = df['rating'].value_counts()[:14] #top 10
sns.barplot(x = bars, y = bars.index)
ax = plt.gca()
for i in ax.patches: 
    plt.text(i.get_width()+0.2, i.get_y()+0.5,  
             str(round((i.get_width()), 2)), 
             fontsize = 10, fontweight ='bold', 
             color ='grey') 
plt.title("Overall Ratings")
plt.xlabel("Count")
```

# Last 10 year trends (Movies & Shows)

```python
# Year wise distribution of TV Shows and Movies Top 10 years

recent_10_year = df[df.release_year > 2011]
g = sns.catplot(
    data=recent_10_year, kind="count",
    x="release_year", hue="type",
    ci="sd", palette="dark", alpha=.6, height=6)
ax = plt.gca()

# ylim max value to be set
y_max = recent_10_year['type'].value_counts().max() 
ax.set_ylim([0, 800])
# Iterate through the list of axes' patches
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
            fontsize=12, color='black', ha='center', va='bottom')
plt.title('Last 10 years trends (Movies and Shows) ',size='15')
plt.show()
```
