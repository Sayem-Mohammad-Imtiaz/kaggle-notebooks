> <h2 style="text-align:center;color:green;font-family:Monospace;"> Model Selection </h2>

```python
X = train_data.drop('price_range',axis=1)
y = train_data['price_range']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

knn= KNeighborsClassifier(n_neighbors=10)
dt= DecisionTreeClassifier(max_depth=10,random_state=0)
svc = SVC()
lr= LogisticRegression()
```

```python
models = [knn,dt,svc,lr]
scores = []

for model in models:
    model.fit(X_train,y_train)
    score = model.score(X_test,y_test)
    scores.append(score)
    print("The score for " + str(model)+" : " + str(score))
```

```python
classifiers = ["KNN","DT","SVC","LR"]
fig = px.bar(scores,color=classifiers,title="Model Evaluation",color_discrete_sequence=px.colors.qualitative.Antique)
fig.show()
```

<p style="color:red;font-family:Monospace;"> From the above graph we can conclude that Logistic  Regression has highest score i.e. 0.962 . </p>


> <h2 style="text-align:center;color:green;font-family:Monospace;">Confusion Matrix  </h2>

```python
from yellowbrick.classifier import ConfusionMatrix

for model in models:
    cm = ConfusionMatrix(
    model, classes=[0,1,2,3])
    cm.fit(X_train, y_train)
    cm.score(X_test, y_test)
    cm.poof()
    
```

> <h2 style="text-align:center;color:green;font-family:Monospace;">Classification Report  </h2>

```python
from sklearn.metrics import classification_report

for model in models:
    print("Classification Report of "+ str(model))
    print("\n")
    print(classification_report(model.predict(X_test), y_test))
    print("\n")
```
