```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df= pd.read_csv("/kaggle/input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv")
df.head()
```

```python
df.dtypes
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(axis=1,inplace=True)
```

```python
X =df
X=X.drop("stroke",axis=1)
y=df.stroke
```

```python
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid= train_test_split(X,y,train_size=0.8,random_state=0)
```

```python
obj_col = [col for col in X_train.columns if X_train[col].dtypes == "object" ]
X_label_train = X_train.copy()
X_label_valid =X_valid.copy()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encode=LabelEncoder()
for col in obj_col:
    X_label_train[col]=label_encode.fit_transform(X_train[col])
    X_label_valid[col]=label_encode.transform(X_valid[col])
```

```python
plt.scatter(X_label_train.smoking_status,y_train)
plt.show()
```

```python
X_label_train.smoking_status
```

```python
from sklearn import svm
clf = svm.SVC(kernel='rbf')
clf.fit(X_label_train, y_train) 
```

```python
SVM_predict= clf.predict(X_label_valid)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(y_valid, SVM_predict)
acc
```
