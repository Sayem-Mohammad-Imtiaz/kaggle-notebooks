```python
data = pd.read_csv('/kaggle/input/cern-electron-collision-data/dielectron.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
data.dropna(axis=0,subset=['M'],inplace=True)
```

```python
data.info()
```

```python
y = data.M
```

```python
y.dtype
```

```python
data.drop(['M'],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_valid,y_train,y_valid = train_test_split(data,y,train_size=0.8,test_size=0.2,random_state=0)
```

```python
from xgboost import XGBRegressor

```

```python
model = XGBRegressor()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
predict = model.predict(X_valid)
print(mean_absolute_error(predict,y_valid))
```
