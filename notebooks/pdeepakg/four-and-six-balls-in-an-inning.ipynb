```python
import plotly.graph_objects as go
b_by_b = pd.read_csv('../input/ipl-complete-dataset-20082020/IPL Ball-by-Ball 2008-2020.csv')
```

```python
# Four & Six hit overs
fig = go.Figure()

df=b_by_b.query("id == 335982 and inning ==1")
df=df.sort_values(by=['over','ball'])
df['score'] = np.cumsum(df.total_runs)
df.index = list(range(len(df.index)))

a=np.zeros(len(df.index), dtype=int)
fours = df.query("batsman_runs == 4").index.astype(int)
sixes = df.query("batsman_runs == 6").index.astype(int)

color_indices_4 = []
color_indices_6 = []

for i in range(len(a)):
    if i in list(fours):
        if i < 5:
            color_indices_4.append(np.round(i/10,1))
            color_indices_6.append(np.nan)
        else:
            color_indices_4.append(np.round(i/6,1))
            color_indices_6.append(np.nan)
    elif i in list(sixes):
        if i < 5:
            color_indices_6.append(np.round(i/10,1))
            color_indices_4.append(np.nan)
        else:
            color_indices_6.append(np.round(i/6,1))
            color_indices_4.append(np.nan)
    else:
        color_indices_4.append(np.nan)
        color_indices_6.append(np.nan)

x = color_indices_4
y = list(df.score.values)

fig.add_trace(go.Scatter(x=x, y=y, mode='markers', name='Four'))

x = list(color_indices_6)
y = list(df.score.values)

fig.add_trace(go.Scatter(x=x, y=y, mode='markers', name='Six'))

fig.update_layout(
    title="Four and Six hit balls in an inning",
    xaxis_title="Overs",
    yaxis_title="Score",
    legend_title="Biggies")

fig.show()
```
