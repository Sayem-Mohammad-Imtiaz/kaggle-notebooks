```python
data = pd.read_csv("/kaggle/input/pokemon-challenge/pokemon.csv")
print(data)
```

```python
#corrolation map
f,ax = plt.subplots(figsize=(13,13))
sns.heatmap(data.corr(), annot=True ,linewidths=.5, fmt='.1f', ax=ax)
plt.show()
data.corr()
```

```python
data.Speed.plot(kind='line',color='green',label="speed", linewidth=1, grid=True , linestyle="-.")
data.Defense.plot(kind="line",color="red",label="defense", linewidth=1, grid=True, linestyle=":")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Line_Plot")
plt.legend()
plt.show()
```

```python
#scatter plot x=attack ; y=defense
plt.scatter(data.Attack , data.Defense,  color="red",   alpha=0.5)
plt.xlabel("Attack")
plt.ylabel("Defense")
plt.title("scatter plot")
plt.show()
```

```python
#histogram
data.Speed.plot(kind="hist", bins=30, color="red" , grid=True ,figsize=(10,10) )
plt.xlabel("Speed")
plt.show()
```
