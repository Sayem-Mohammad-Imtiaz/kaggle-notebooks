PIVOTING DATA FRAMES

```python
dic={"treatment":["A","A","B","B"],
    "gender":["F","M","F","M"],
    "response":[10,45,5,9],
    "age":[15,4,72,65]}
df=pd.DataFrame(dic)
df
```

```python
df.pivot(index="treatment" , columns="gender" , values="response")
```

STACKING UND UNSTACKING DATA FRAMES

```python
df1=df.set_index(["treatment" , "gender"])
df1
```

```python
df1.unstack(level=0)
```

```python
df1.unstack(level=1)
```

```python
df2=df1.swaplevel(0,1)
df2
```

MELTING DATA FRAMES

```python
pd.melt(frame=df , id_vars="treatment" , value_vars=["age" , "response"])
```
