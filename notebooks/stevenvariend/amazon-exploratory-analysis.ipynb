# <font color='blue'>Amazon exploratory analysis </font> 

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
allbooks = pd.read_csv('../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
```

# <font color='blue'>Duplicates found </font> 

```python
allbooks.drop_duplicates("Name",inplace=True)
```

```python
topfifty = allbooks.sort_values(["User Rating","Reviews"],ascending = False).head(50)
topfifty
```

```python
topfifty.groupby("Year").count().sort_values("Name")["Name"]
```

```python
topfifty.groupby("Author").count().sort_values("Name")["Name"]
```

# <font color='blue'>Harmonizing J. K. Rowling</font> 

```python
allbooks["Author"].replace("J. K. Rowling","J.K. Rowling",inplace = True)

topfifty = allbooks.sort_values(["User Rating","Reviews"],ascending = False).head(50)

topfifty.groupby("Author").count().sort_values("Name")["Name"]
```

# <font color='blue'>Visualisation - allbooks</font> 

```python
plt.figure(figsize=(6,3),dpi=150)
sns.countplot(x='Genre',data=allbooks, palette = "magma");
```

```python
plt.figure(figsize=(6,3),dpi=150)
sns.scatterplot(data=allbooks,x="Reviews",y="User Rating",alpha=0.2);
```

```python
plt.figure(figsize=(6,3))
sns.jointplot(data=allbooks,x="Reviews",y="User Rating",hue="Genre",alpha=0.2,height=8);
```

```python
plt.figure(figsize=(6,3),dpi=150)
sns.kdeplot(data=allbooks,x="Price", fill = True);
```

```python
plt.figure(figsize=(6,3),dpi=150)
sns.kdeplot(data=allbooks,x="User Rating", fill = True);
```

```python
plt.figure(figsize=(6,3),dpi=150)
sns.kdeplot(data=allbooks,x="Price",hue="Genre",fill=True,palette="Dark2");
```

# <font color='blue'>Visualisation - topfifty</font> 

```python
plt.figure(figsize=(6,3),dpi=150)
sns.countplot(x='Genre',data=topfifty, palette = "magma");
```
