```python
# Making  the independent variables matrix, Annual and Spending score
X = df.iloc[:, [3, 4]].values
```

```python
# Convert gender to categorical variabel
df = pd.get_dummies(df, columns = ['Gender'], prefix = ['Gender'])
df.head(3)
```

> KMEANS

```python
#Using KMeans for clustering Annual Income and Spending Score
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
font_title = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 35}

font_axes = {'family' : 'normal',
        'weight' : 'normal',
        'size'   : 28}
```

```python
#Plotting Number of Clusters Vs wcss - The Elbow Method Annual Income
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method', fontsize=20)
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

Based on the Elbow Method graph above, the best number of clusters is 5. So it will be analyzed how the clustering between Annual Income and Spending Score with 5 clusters uses the K-Means algorithm.

```python
#Taking number of clusters = 5
kmeans = KMeans(n_clusters = 5, init = 'k-means++', max_iter = 300, n_init = 10)
y_kmeans = kmeans.fit_predict(X)
```

```python
# PLotting the clusters
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'yellow', label = 'Cluster4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'pink', label = 'Cluster5')
plt.title('Clusters of Customers', **font_title)
plt.xlabel('Annual income(k$)', **font_axes)
plt.ylabel('spending score', **font_axes)
plt.legend()
plt.show()
```

**Result:**
1. Cluster 1 (red color), is called the waster cluster

In cluster 1 it can be seen that people with low income but have a higher spending score, they are people who for some reason prefer to buy products even though they have low income. Maybe because these people are more than satisfied with the mall's services. Shops/malls may not target these people effectively but still won't miss them.
2. Cluster 2 (blue color), is called the common cluster

In cluster 2, it can be seen that mall customers who have moderate income have moderate expenses, adjusted to their economic situation.

3. Cluster 3 (green), called the stingy cluster

In cluster 3 it can be seen that high-income mall customers have low spending, this is interesting. Maybe these are people who are dissatisfied or unhappy with the mall's services. This can be a prime target for malls, as it has the potential to cost money. Therefore, the mall manager will try to add new facilities in order to attract the interest of the community and meet their needs.

4. Cluster 4 (yellow color), is called a conglomerate cluster

In cluster 4 it can be seen that people have high income and high expenses, this is an ideal case for malls or shops because these people are the main source of profit. These people may be regulars of the mall and are confident in the mall's amenities.

5.  Cluster 5 (pink color), is called the caution cluster

In cluster 5 it can be seen that people have low annual income and low spending scores, this is quite reasonable because people who have low salaries prefer to buy less, in fact, these are wise people who know how spend and save money. Shops/malls will be least interested in people belonging to this cluster.

```python
# Making  the independent variables matrix (Age)
X2 = df.iloc[:, [1, 3]].values
```

```python
#Using KMeans for clustering (Age)
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10)
    kmeans.fit(X2)
    wcss.append(kmeans.inertia_)
```

```python
#Plotting Number of Clusters Vs wcss - The Elbow Method (Age)
plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method', fontsize=20)
plt.xlabel('Number of clusters')
plt.ylabel('wcss')
plt.show()
```

The best number of clusters is 4. So it will be analyzed how the clustering between Age and Spending Score with 4 clusters using the K-Means algorithm.

```python
#Taking number of clusters = 4
kmeans = KMeans(n_clusters = 4, init = 'k-means++', max_iter = 300, n_init = 10)
y_kmeans = kmeans.fit_predict(X2)

# PLotting the clusters
plt.scatter(X2[y_kmeans == 0, 0], X2[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster1')
plt.scatter(X2[y_kmeans == 1, 0], X2[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster2')
plt.scatter(X2[y_kmeans == 2, 0], X2[y_kmeans == 2, 1], s = 100, c = 'green', label = 'Cluster3')
plt.scatter(X2[y_kmeans == 3, 0], X2[y_kmeans == 3, 1], s = 100, c = 'yellow', label = 'Cluster4')

plt.title('Clusters of Age', **font_title)
plt.xlabel('Age', **font_axes)
plt.ylabel('spending score', **font_axes)
plt.legend()
plt.show()
```

**Result:**
1. Cluster 1 (red color), is called the young customer cluster

In cluster 1, customers with an age range of 20–40 years have a spending score of 30–70. In this cluster, mall owners need to improve their services.

2. Cluster 2 (blue), called priority customer cluster

In cluster 2, customers with an age range of 20–40 years have a spending score of 70–99. In this cluster, mall owners need to maintain good relations while maintaining mall services.

3. Cluster 3 (green color), called old customer cluster cluster

In cluster 3, customers with an age range of 40–70 years have a spending score of 35–60. In this cluster, mall owners need to improve related facilities to make it easier for adults and even the elderly to shop at the mall to feel comfortable so that they can increase their shopping activities.

4. Cluster 4 (yellow color), is called the target customer cluster

In cluster 4, customers with an age range of 20–70 years have a spending score of 0–30. In this cluster, mall owners need to pay more attention to improving mall services and facilities.
