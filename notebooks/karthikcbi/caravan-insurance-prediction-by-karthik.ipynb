We will be considering all such columns with less than 5 unique values as categorical and any column with 5 or more unique values as numerical

```python
cat_cols = df_train.columns[df_train.nunique() < 5]
```

```python
cat_columns = cat_cols.tolist()
```

```python
num_cols = df_train.columns[df_train.nunique() >= 5]
```

```python
import matplotlib.pyplot as plt    
import seaborn as sns
```

```python
plt.figure(figsize=(15,18))
for i,j in enumerate(df_train[num_columns]):
   plt.subplot(9,7,i+1)
   sns.boxplot(y = df_train[j])
```

It is evident from the above box plots that many of the numerical columns have outliers, hence we will consider Robust Scalar for numeric columns and One Hot Encoding for Cat Columns

```python
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer([
                        ('abc', RobustScaler(),num_columns),
                        ('cde', OneHotEncoder(), cat_columns)    
                        ],
                        remainder = 'passthrough'
                        )
```

```python
X_train = df_train
X_test = df_test
```

```python
from sklearn.pipeline import Pipeline
```

```python
pipe = Pipeline([
                ('ct', ct),
                ('rf', rf())
                ])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```
