```python
# importing important libraries
import pandas as pd
import numpy as np # data pre-processing

import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.style.use('dark_background')
import plotly.express as px # data visualization

import warnings
warnings.filterwarnings("ignore") # additional
```

```python
df = pd.read_csv('../input/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv') # loading the data
```

```python
df =df.replace('Dadra and Nagar Haveli and Daman and Diu', 'Daman and Diu')
```

<h1 style="background-color:powderblue;"><center><b>Data Visualization</b></center></h1>

```python
plt.figure(figsize=(15,10))
sub = np.triu(df.corr())
sns.heatmap(df.corr(), annot=True,linewidth=.15, mask=sub)
```

* Note the correlation between variables.
* The **more** the **Total Cases** the **more** the **Discharge count**.
* **Total Cases** is also directly related to **Deaths** count

```python
px.bar(df, x="State/UTs", y="Total Cases", color="Death Ratio", title="Total Cases as per each State : ")
```

* Although **Maharastra** is having **highest total cases** count, it does **not** have **highest death ratio** count.
* **Punjab** has **highest Death Ratio** despite having very less total cases count compared to other countries.
* **Maharastra** is having **better medical facilities than Punjab**.
* Likewise we can compare other states based on their total cases and death ratio count.

```python
px.bar(df, x="State/UTs", y="Active", color="Active Ratio", title="Active Cases as per each State : ")
```

* **Maharastra** is having **highest Active** count.
* But **Mizoram** is having **highest Active Ratio** count !
* We cannot say the **people** are **most likely** to **get infected** by corona virus in Maharastra as despite of having more Active cases, Maharastra is having less **Active Ratio** count. 
* **Covid** is spreading **most rapidly** in **Mizoram**now as compared to any other state.

```python
px.scatter(df, x='Active Ratio',y='Death Ratio', color=df['State/UTs'])
```

* Again we can see that **Punjab** is the state which requires most **attention** to the **medical facilities** as it has **highest** value for **Death Ratio**.
* **Mizoram** is the state which has **highest Active Ratio**.
* **Sikkim** is the state with pretty **high death** as well as **Active ratio**.

```python
px.scatter(df, x="Total Cases", y="Active",size="Active Ratio", color="Active Ratio",hover_name="State/UTs", log_x=True, size_max=60)
```

* Although **Maharastra** is having **more** number of **Active** and **Total Cases** it **does not** have **highest Active Ratio**.
* **Mizoram** is having **highest Active Ratio** despite of having **very less Total Cases and Active** count.
* **Lockdown** and **precautive measures** should be taken considering all these factors.
