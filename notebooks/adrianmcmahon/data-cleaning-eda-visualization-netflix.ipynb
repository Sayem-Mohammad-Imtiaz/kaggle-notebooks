# Cleaning the Data.

```python
dummies = df['Genre'].str.get_dummies(sep='/')
df_genre = pd.concat([df, dummies], axis=1)

merge_genre = df_genre['Action'] + df_genre['Action comedy'] + df_genre['Historical-epic'] + df_genre['Action thriller'] + df_genre['Action-adventure'] + df_genre['Action-thriller']
df_genre['Action1'] = merge_genre
df_genre['Action1'] = df_genre.Action1.replace((2), 1)

merge_genre = df_genre['Adventure'] + df_genre['Adventure-romance'] + df_genre['Action-adventure']
df_genre['Adventure1'] = merge_genre
df_genre['Adventure1'] = df_genre.Adventure1.replace((2), 1)

merge_genre = df_genre['Animation'] + df_genre['Animated musical comedy'] + df_genre['Animation'] + df_genre['Anime'] + df_genre['Animation '] + df_genre['Anime ']
df_genre['Animation1'] = merge_genre
df_genre['Animation1'] = df_genre.Animation1.replace((2), 1)

merge_genre = df_genre['Comedy'] + df_genre['Action comedy'] + df_genre['Animated musical comedy'] + df_genre['Black comedy'] + df_genre['Christmas comedy'] + df_genre['Comedy horror'] + df_genre['Comedy mystery'] + df_genre['Comedy-drama'] + df_genre['Coming-of-age comedy-drama'] + df_genre['Dance comedy'] + df_genre['Dark comedy'] + df_genre['Superhero-Comedy'] + df_genre['Teen comedy horror'] + df_genre['Teen comedy-drama'] + df_genre['War-Comedy'] + df_genre['Drama-Comedy'] + df_genre['Hidden-camera prank comedy'] + df_genre['Horror comedy'] + df_genre['Musical comedy'] + df_genre['Romantic comedy'] + df_genre['Romantic comedy-drama'] + df_genre[' Comedy'] + df_genre['Comedy ']
df_genre['Comedy1'] = merge_genre
df_genre['Comedy1'] = df_genre.Comedy1.replace((2), 1)

merge_genre = df_genre['Musical'] + df_genre['Dance comedy'] + df_genre['Animated musical comedy'] + df_genre['Christian musical'] + df_genre['Christmas musical'] + df_genre['Musical comedy'] + df_genre[' Musical'] + df_genre[' Musicial'] + df_genre['Musical ']
df_genre['Musical1'] = merge_genre
df_genre['Musical1'] = df_genre.Musical1.replace((2), 1)

merge_genre = df_genre['Action thriller'] + df_genre['Political thriller'] + df_genre['Action-thriller'] + df_genre['Crime thriller'] + df_genre['Horror thriller'] + df_genre['Horror-thriller'] + df_genre['Psychological thriller'] + df_genre['Psychological thriller drama'] + df_genre['Romantic thriller'] + df_genre['Science fiction thriller'] + df_genre['Spy thriller'] + df_genre['Thriller'] 
df_genre['Thriller1'] = merge_genre
df_genre['Thriller1'] = df_genre.Thriller1.replace((2), 1)

merge_genre = df_genre['Comedy-drama'] + df_genre['Coming-of-age comedy-drama'] + df_genre['Drama'] + df_genre['Drama '] + df_genre['Drama-Comedy'] + df_genre['Psychological thriller drama'] + df_genre['Romance drama'] + df_genre['Romantic comedy-drama'] + df_genre['Romantic drama'] + df_genre['Romantic teen drama'] + df_genre['Romantic teenage drama'] + df_genre['Sports-drama'] + df_genre['Supernatural drama'] + df_genre['Teen comedy-drama'] +df_genre['War drama']
df_genre['Drama1'] = merge_genre
df_genre['Drama1'] = df_genre.Drama1.replace((2), 1)

merge_genre = df_genre[' Science Fiction'] + df_genre['Fantasy'] + df_genre['Satire'] + df_genre['Science fiction'] + df_genre['Science fiction adventure'] + df_genre['Science fiction thriller'] + df_genre['Superhero'] + df_genre['Superhero-Comedy'] + df_genre['Supernatural drama'] + df_genre['Urban fantasy'] + df_genre['Zombie']
df_genre['Science Fiction1'] = merge_genre
df_genre['Science Fiction1'] = df_genre['Science Fiction1'].replace((2), 1)

merge_genre = df_genre['Crime drama'] + df_genre['Heist'] + df_genre['Heist film'] + df_genre['Crime thriller'] + df_genre['Crime thriller'] + df_genre['Mystery'] + df_genre['Aftershow ']
df_genre['Crime1'] = merge_genre
df_genre['Crime1'] = df_genre.Crime1.replace((2), 1)

merge_genre = df_genre['Adventure-romance'] + df_genre['Romance'] + df_genre['Romance drama'] + df_genre['Romantic comedy'] + df_genre['Romantic comedy-drama'] + df_genre['Romantic drama'] + df_genre['Romantic teen drama'] + df_genre['Romantic teenage drama'] + df_genre['Romantic thriller']
df_genre['Romance1'] = merge_genre
df_genre['Romance1'] = df_genre.Romance1.replace((2), 1)

merge_genre = df_genre['Comedy horror'] + df_genre['Horror'] + df_genre['Horror anthology'] + df_genre['Horror comedy'] + df_genre['Horror thriller'] + df_genre['Horror-thriller'] + df_genre['Psychological horror'] + df_genre['Teen comedy horror'] + df_genre['Romantic thriller']
df_genre['Horror1'] = merge_genre
df_genre['Horror1'] = df_genre.Horror1.replace((2), 1)

merge_genre = df_genre['Family'] + df_genre['Family film']
df_genre['Family Flim1'] = merge_genre
df_genre['Family Flim1'] = df_genre['Family Flim1'].replace((2), 1)

merge_genre = df_genre['Variety show'] + df_genre['Variety Show']
df_genre['Variety Show1'] = merge_genre
df_genre['Variety Show1'] = df_genre['Variety Show1'].replace((2), 1)

merge_genre = df_genre['War'] + df_genre['War-Comedy']
df_genre['War1'] = merge_genre
df_genre['War1'] = df_genre.War1.replace((2), 1)

df_genre = df_genre.drop(['Genre', 'Action', 'Family', 'Holiday', 'War', 'War-Comedy', 'Variety Show', 'Variety show', 'Historical drama', 'Historical-epic', 'Heist', 'Heist film', 'Family film', 'Crime drama', 'Christmas', 'Anthology', 'Biographical', ' Interview', ' Short', 'Aftershow ', 'Political thriller', 'Comedy horror', 'Horror', 'Horror anthology', 'Mystery', 'Horror comedy', 'Adventure-romance', 'Horror thriller', 'Crime thriller', 'Horror-thriller', 'Romance', 'Psychological horror', 'Romance drama', 'Teen comedy horror', 'Romantic comedy', 'Fantasy', 'Dance comedy', 'Romantic comedy-drama', 'Romantic drama', 'Comedy mystery', 'Romantic teen drama', 'Romantic teenage drama', 'Dark comedy', 'Romantic thriller', 'Adventure', 'Animation', 'Action-adventure', 'Anime', 'Action comedy', 'Black comedy', 'Christmas comedy', 'Hidden-camera prank comedy', 'Comedy', 'Animated musical comedy', 'Christian musical', 'Musical', 'Christmas musical', 'Musical comedy', 'Action thriller', 'Action-thriller', 'Psychological thriller', 'Spy thriller', 'Thriller', ' Comedy', 'Comedy ', ' Musical', ' Musicial', 'Animation ', 'Anime ', 'Musical ', 'Comedy-drama', 'Coming-of-age comedy-drama', 'Drama', 'Drama ', 'Drama-Comedy', 'Psychological thriller drama', 'Sports-drama', 'Teen comedy-drama', 'War drama', 'Satire', 'Science fiction', 'Science fiction adventure', 'Science fiction thriller', ' Science Fiction', 'Superhero', 'Superhero-Comedy', 'Supernatural drama', 'Urban fantasy', 'Zombie'], axis=1)
df_genre = df_genre.rename(columns = {'Action1': 'Action', 'Adventure1': 'Adventure', 'Animation1': 'Animation', 'Comedy1': 'Comedy', 'Musical1': 'Musical', 'Thriller1': 'Thriller', 'Drama1': 'Drama', 'Science Fiction1': 'Science Fiction', 'Crime1': 'Crime', 'Romance1': 'Romance', 'Horror1': 'Horror', 'Family Flim1': 'Family Flim', 'Variety Show1': 'Variety Show', 'War1': 'War'})
```

### Language cleaning

```python
dummies = df['Language'].str.get_dummies(sep='/')
df_language = pd.concat([df, dummies], axis=1)

merge_language = df_language['Thai'] + df_language['Thia']
df_language['Thai1'] = merge_language
df_language['Thai1'] = df_language.Thai1.replace((2), 1)

df_language = df_language.drop(['Language', 'Thai', 'Thia'], axis=1)
df_language = df_language.rename(columns = {'Thai1': 'Thai'})
```

### Date cleaning

```python
df_date = df.copy()
df_date['Day of Week'] = df_date['Premiere'].dt.day_name()
df_date['Year'] = df_date['Premiere'].dt.year
df_date['Month'] = df_date['Premiere'].dt.month
```

# Plotting the data

```python
df_genre_count = df_genre.drop(['Title', 'Premiere', 'Runtime', 'IMDB Score', 'Language'], axis=1)
df_genre_count = df_genre_count.sum().sort_values(ascending = False)
df_genre_count = df_genre_count.reset_index()
df_genre_count.columns = ['Genre', 'Count']
```

```python
month_count = df_date.groupby('Month').apply(lambda x:x['Title'].count()).reset_index(name='Count')
month_count.Month = month_count.Month.replace(1, 'January').replace(2, 'February').replace(3, 'March').replace(4, 'April').replace(5, 'May').replace(6, 'June').replace(7, 'July').replace(8, 'August').replace(9, 'September').replace(10, 'October').replace(11, 'November').replace(12, 'December')

fig = px.bar(
                month_count, 
                y='Count', 
                x='Month')
fig.update_layout(
                title='Number of Movies Release according to Month',
                titlefont_size = 28, 
                xaxis = dict (
                    title = 'Month of Release', 
                    titlefont_size=16), 
                yaxis=dict(
                    title='Count of Movies', 
                    titlefont_size=16))
fig.show()
```

#### The month of April and October are when maximum movies are released. They are exactly 6 months apart.
