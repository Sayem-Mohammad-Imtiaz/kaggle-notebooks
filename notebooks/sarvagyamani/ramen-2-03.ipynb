```python
ramen_ratings=pd.read_csv('/kaggle/input/ramen-ratings/ramen-ratings.csv')

ramen_ratings
```

```python
ramen_ratings.info()
```

```python
ramen_ratings[ramen_ratings['Stars']== 'Unrated'] = np.nan
ramen_ratings['Stars'] = ramen_ratings['Stars'].astype(float)
```

```python
avg_rate = ramen_ratings['Stars'].mean()
avg_rate
```

```python
ramen_ratings[ramen_ratings['Stars']== np.nan] = avg_rate
ramen_ratings['Stars'].fillna(avg_rate,inplace = True)
ramen_ratings.isnull().sum()
```

```python
rated_restaurants = ramen_ratings.dropna()
ramen_ratings.drop('Top Ten',axis=1,inplace=True)
```

```python
ramen_ratings.dropna(axis=0,how='any',inplace=True)
ramen_ratings.isnull().sum()
```

```python
country=ramen_ratings.Country.tolist()

Review={}
for cname in country:
    if cname not in Review:
        Review[cname]=1
        
    else:
        Review[cname]+=1

print('maximum rating from : '+str(sorted(Review,key = lambda r : Review[r])[-1])+' : '+str(max(Review.values())))
print('minimum rating from : '+str(sorted(Review,key = lambda r : Review[r])[0])+' : '+str(min(Review.values())))
```

# So the most Reviews are from **Japan:352** and the Least Reviews are from **USA and Nigeria:1**

```python
brand=ramen_ratings.Brand.tolist()
Brand_ratings={}
for bname in brand:
    if bname not in Brand_ratings:
        Brand_ratings[bname]=1
    else:
        Brand_ratings[bname]+=1
        
one_rating_list=[]
for rating in Brand_ratings:
    if Brand_ratings[rating]==1:
        one_rating_list.append(rating)

print("Brands with 1 Star Rating: {}".format(len(one_rating_list)))
print('maximum rating : '+sorted(Brand_ratings, key= lambda x: Brand_ratings[x])[-1])
print('minimum rating : '+sorted(Brand_ratings, key= lambda x: Brand_ratings[x])[0])

```
