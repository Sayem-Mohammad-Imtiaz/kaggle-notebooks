## Modelling

```python
X=df.iloc[:, 1:3].values
y=df.iloc[:, 0].values

#change the position of the target column to the left
df = df[['Height', 'Weight', 'Gender']]
df.head()
```

## Splitting dataset into training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.25, random_state=0)
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

For this prediction I only use 3 classifier, namely Decision Tree, KNN, and Random Forest. There are so many classification algorithm, but for now I only use these three calssifiers. 

## Prediction using Decision Tree

```python
from sklearn import tree
clf1 = tree.DecisionTreeClassifier()
clf1.fit(X_train, y_train)
```

```python
prediction1 = clf1.predict(X_test)
r1 = accuracy_score(y_test, prediction1)
print('Accuracy of Decision Tree =',r1)
```

## Prediction using KNN

```python
from sklearn import neighbors
clf2 = neighbors.KNeighborsClassifier(n_neighbors=17)
clf2.fit(X_train, y_train)
```

```python
prediction2 = clf2.predict(X_test)
r2 = accuracy_score(y_test, prediction2)
print('Accuracy of Decision Tree =',r2)
```

## Prediction using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
clf3 = RandomForestClassifier(max_depth=4)
clf3.fit(X_train, y_train)
```

```python
prediction3 = clf3.predict(X_test)
r3 = accuracy_score(y_test, prediction3)
print('Accuracy of Decision Tree =',r3)
```

From the modeling result we use the DecisionTreeClassifier to predict the gender because of the better accuracy. You can change the value of the height and weight.

```python
height = 71
weight = 205
pred = clf2.predict([[height, weight]])
print('Your gender is', pred)
```
