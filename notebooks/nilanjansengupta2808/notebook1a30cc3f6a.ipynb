```python
record = pd.read_csv('/kaggle/input/covid19-in-india/covid_19_india.csv', parse_dates=['Date'], dayfirst=True)
record.head()
```

```python
#Data cleaning keeping relevant coloumns and renaming them for better understanding
record= record[['Date', 'State/UnionTerritory','Cured','Deaths','Confirmed']]
record.columns = ['date', 'state','cured','deaths','confirmed']
```
