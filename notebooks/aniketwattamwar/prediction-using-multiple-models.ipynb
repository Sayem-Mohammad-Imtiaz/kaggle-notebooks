```python
#reading the data
data = pd.read_csv('/kaggle/input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv')
data.head()
```

```python
# separating the training and output data
train = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
data.isna().sum()
```

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
train = sc_X.fit_transform(train)
```

# Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(train, y, test_size = 0.20, random_state = 42)
```

# Training on Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(max_depth=15, random_state=0)
clf.fit(X_train, y_train)
```

**Prediction with RF**

```python
pred = clf.predict(X_test)
pred
```

**Checking the accuracy score**

```python
from sklearn.metrics import accuracy_score
acc_rf = accuracy_score(y_test,pred)
acc_rf
```

# Training on XGBoost Classifier

```python
from xgboost import XGBClassifier
model = XGBClassifier(learning_rate =0.04,n_estimators=500)
model.fit(X_train, y_train)
```

**Prediction using xgboost**

```python
y_pred_xg = model.predict(X_test)
y_pred_xg
```

**Accuracy using xgboost**

```python
acc_rf = accuracy_score(y_test,y_pred_xg)
acc_rf
```
