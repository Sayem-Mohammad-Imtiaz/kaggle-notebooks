```python
#using OneHotEncoding for the categorical column which has more than 2 category.

categorical_column = ['MultipleLines','InternetService','OnlineSecurity',
                      'OnlineBackup','DeviceProtection','TechSupport',
                      'StreamingTV','StreamingMovies','Contract','PaymentMethod']
```

```python
final_df = pd.get_dummies(columns=categorical_column,data = df)
```

```python
final_df.info()
```

```python
#As per the above information TotalCharges is of object type
final_df['TotalCharges'] = pd.to_numeric(final_df.TotalCharges,errors='coerce')
final_df['TotalCharges'].fillna(final_df['TotalCharges'].mean(),inplace = True)
```

```python
final_df.info()
```

```python
X = final_df.drop('Churn',axis=1)
Y = final_df['Churn']
#split the data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=0)
X_train.shape,X_test.shape,Y_train.shape,Y_test.shape
```

```python
#Apply logistic_regression
lr = LogisticRegression()
model = lr.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
print(pd.DataFrame(confusion_matrix(Y_test,Y_pred)))
```

```python
print(classification_report(Y_test,Y_pred))
```

```python
#Applying RandomForest
rf = RandomForestClassifier(n_estimators=1000,max_depth=2,random_state=41)
model1 = rf.fit(X_train,Y_train)
```

```python
y_rf_pre = model1.predict(X_test)
```

```python
print(pd.DataFrame(confusion_matrix(Y_test,y_rf_pre)))
```

```python
print(classification_report(Y_test,y_rf_pre))
```
