```python
df=pd.read_csv('/kaggle/input/voicegender/voice.csv')
df.info()
df.describe()
```

```python
df
```

```python
df=df.drop(['label'],axis = 1)
df=df.dropna()
df
```

```python
standard_df = StandardScaler().fit_transform(df)
standard_df.shape
pca=decomposition.PCA()
pca.n_components=3
pca_df=pca.fit_transform(df)
df=pd.DataFrame(pca_df)
df
```

```python
clusters=range(1,10)
inertias=[]
for i in clusters:
    kmean=KMeans(n_clusters=i,init='k-means++',random_state=1)
    kmean.fit(df)
    inertias.append(kmean.inertia_)
    print(f'{i} calculated')
    
plt.figure(figsize=(12,5))
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.plot(clusters,inertias);
```

```python
df2=df
df3=df
df4=df
kmeans2=KMeans(n_clusters=2,init='k-means++',random_state=1).fit(df2)
kmeans3=KMeans(n_clusters=3,init='k-means++',random_state=1).fit(df3)
kmeans4=KMeans(n_clusters=4,init='k-means++',random_state=1).fit(df4)
labels2=kmeans2.labels_
print('labels2:',labels2)
labels3=kmeans3.labels_
print('labels3:',labels3)
labels4=kmeans4.labels_
print('labels4:',labels4)
```

```python
df2['Cluster']=pd.Series(labels2)
cluster_names2={0:'Cluster 0',
              1:'Cluster 1'}
df2['Cluster_names']=df2['Cluster'].map(cluster_names2)
scatter=px.scatter_3d(x=0,y=1,z=2,data_frame=df2,color=df2['Cluster_names'])
scatter.show()
```

```python
df3['Cluster']=pd.Series(labels3)
cluster_names3={0:'Cluster 0',
              1:'Cluster 1',
              2:'Cluster 2'}
df3['Cluster_names']=df3['Cluster'].map(cluster_names3)
scatter=px.scatter_3d(x=0,y=1,z=2,data_frame=df3,color=df3['Cluster_names'])
scatter.show()
```

```python
df4['Cluster']=pd.Series(labels4)
cluster_names4={0:'Cluster 0',
              1:'Cluster 1',
              2:'Cluster 2',
               3:'Cluster 3'}
df4['Cluster_names']=df4['Cluster'].map(cluster_names4)
scatter=px.scatter_3d(x=0,y=1,z=2,data_frame=df4,color=df4['Cluster_names'])
scatter.show()
```
