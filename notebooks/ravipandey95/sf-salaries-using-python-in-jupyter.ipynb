```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

Read Salaries.csv as a dataframe called sal

Here we can see that apart from "TotalPay","TotalPayBenefits"and"Notes" none of the columns have float or int datatype so to convert an object datatype to float datatype we can do as follow:

```python
sal['BasePay']=pd.to_numeric(sal.BasePay, errors = 'coerce')
sal['OvertimePay']=pd.to_numeric(sal.OvertimePay, errors = 'coerce')
sal['OtherPay']=pd.to_numeric(sal.OtherPay, errors = 'coerce')
```

** What was the average (mean) BasePay of all employees per year? (2011-2014) ? **

```python
sal.groupby('Year')['BasePay'].mean()


# sns.barplot(x='Year',y='BasePay',data=sal) #use this to plot remove'#' before sns 
```

** What was the average (mean) OvertimePay of all employees per year? (2011-2014) ? **

```python
sal.groupby('Year').mean()['OvertimePay']

#sns.barplot(x='Year',y='OvertimePay',data=sal) #use this to plot remove'#' before sns 
```

** What was the average (mean) TotalPayBenefits of all employees per year? (2011-2014) ? **

```python
sal.groupby('Year')['TotalPayBenefits'].mean()
```

** What are the top 5 most common jobs? **

```python
sal['JobTitle'].value_counts().head(5)
```

** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
sum(sal[sal['Year']==2013]['JobTitle'].value_counts() == 1)
```

** How many people have the word Chief in their job title? **

```python
def chief_string(title):
    if 'chief' in title.lower():
        return True
    else:
        return False
```

```python
sum(sal['JobTitle'].apply(lambda x: chief_string(x)))
```
