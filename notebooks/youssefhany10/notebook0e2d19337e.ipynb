```python
df_train = pd.read_csv('/kaggle/input/mobile-price-classification/train.csv')
df_test = pd.read_csv('/kaggle/input/mobile-price-classification/test.csv')
```

```python
df_train.head()
```

```python
X = df_train.drop(columns=['price_range'])
```

```python
y = df_train['price_range']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
 from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'p':[1,3], 'n_neighbors':[1, 10]}
```

```python
clf = GridSearchCV(knn, parameters)
```

```python
clf = clf.fit(X_train,y_train)
```

```python
clf.best_estimator_
```

```python
knn_new = KNeighborsClassifier(n_neighbors=10, p=3)
```

```python
knn_new = knn_new.fit(X_train,y_train)
y_predict = knn_new.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_predict)
```
