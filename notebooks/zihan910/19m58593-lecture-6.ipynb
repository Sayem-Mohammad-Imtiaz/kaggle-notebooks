```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
np.set_printoptions(threshold=np.inf)

selected_country = 'Italy'
df = pd.read_csv('../input/novel-corona-virus-2019-dataset/covid_19_data.csv', header=0)
#select the dataframe that contains Italy
df = df[df['Country/Region']==selected_country]
df = df.groupby('ObservationDate').sum()
print(df)
```

# Daily Cases Calculation

Here,the time series plot shows the daily confirmed (in blue), daily recovered (in green), and the daily death (in yellow) cases in Italy from 1/31/2020 up to now. 

```python
df['Daily Confrimed'] = df['Confirmed'].diff()
df['Daily Deaths'] = df['Deaths'].diff()
df['Daily Recovered'] = df['Recovered'].diff()

df['Daily Confrimed'].plot()
df['Daily Deaths'].plot()
df['Daily Recovered'].plot()
```

From this time-series plot, it can be clearly observed that the overall trend of the confirmed corona cases is decreasing with time, after it reaches the peak on March 21, 2020, with 6557 confirmed cases in one day. 

From January to March, the daily confirmed cases have been dramatically increased. As the plot shows, the COVID-19 cases are controlled and the cases are gradually decreasing since March. The decreasing trend can also be seen from the increase of the daily recovery (in green).

From January 31st, the government of Italy declared a state of emergency that lasted 6 months (Xinhua Net, 2020). The lockdown of the entire country began on March 10th. After applying two times of quarantine extensions, the final quarantine time was extended to May 3rd. On March 22nd ,2020, the government of Italy has published the decree to suspend all industrial and commercial activities. This suspension measure was effective in all provinces in Italy from March 23rd to April 3rd 2020 (KPMG, 2020).From March to May, the number of confirmed cases was effectively decreased via appropriate corona response measures. Thus, on May 4th, the quarantine time was finally to an end. 

In Italy’s response to COVID-19, the government firstly prioritized the citizens’ primary right of health and well-beings. Although some decision delay occurred in the early stage of the response to COVID-19, the Italian government declared the state of emergency on January 31st, 2020 (Forbes, 2020). In comparison, the initial decision in UK was to focus on the protection of economic growth, instead of prioritizing the citizens’ health. This economic-oriented response to COVID-19 has certain downside in lack of protection of individuals’ health. 

Secondly, Italian government and institutions have put effort in providing the most transparent information of COVID-19 to the residents (Forbes, 2020). During the early stage of the COVD-19 spread in Italy, the government and institutions also worked effectively to control the misleading and false information on social media. The third key point of Italy’s response is the concept of integration approach and flexible strategy. A military-engineer team from the Italian Siare Engineering International Group is incorporated in the company to make sure the 24-hour production of the lung ventilators. In the fashion industry, companies such as Valentino, Caledonia, Geox, and Prada started their own mask production in fighting against COVID-19. 

Besides, the Italian institutions worked effectively to construct the infrastructure needed for aiding the emergency, such as “COVID hospitals” (Forbes, 2020). An integrated emergency transportation system, which combines the ground (the ambulance), and the air (helicopter), has successfully saved the time for saving lives. Italian Civil Protection has called volunteer help for 500 nurses, but over 9500 applications were received within the first 48 hours when the call was released (Forbes, 2020). By applying the quick response to this global pandemic, Italy has fought with COVID-19 in its effective way.
