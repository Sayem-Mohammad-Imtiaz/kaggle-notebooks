## Folium Lib & Choropleth
* ###   In this Notebook we will see the basic methods available in the Folium Library for Maps.
* ####    Folium library is used for visualising WorldMap (cities, countries, continents)
* ####    with specific Lat long coords we can visualise close view of cities, streets, locations.
* ####    We can add custom geo.json files as additional layer to view data stats distributed in the map.

```python
#importing necessary libraries
import pandas as pd #For Dataset import for stats layer
import folium       #Folium library for map visualisation and customisation
import json         #json library for importing geo.json file
```

```python
#Folium map object creation
Worldmap = folium.Map(zoom_start=18)
Worldmap
```

### Customisation:
* #### **Tiles**: This argument when passed in the method with specific tilestyle name: it formats the layout of the map

* #### **Marker**: Adds a marker in the specified latlong that can be added to the map with *'.add_to'(**map**)*

* #### Below eg. for placing location marker at *WallStreet* with popup marker and map vis style name: *'Stamen Toner'*

```python
Ny_m = folium.Map(location=[40.7128, -74.0060], zoom_start=12, tiles = "Stamen Toner")

tooltip = "Click me!"

folium.Marker(
    [40.706005, -74.008827], popup="<i>Wall Street</i>", tooltip=tooltip
).add_to(Ny_m)

Ny_m
```

### **Displaying stats with .geojson file overlay**:
* ####   import the dataset containing the stats - *Country Name* should be available in the data for choropleth to index

* ####   Below eg. we will update Data stats layer with *'World Happiness report - 2021'* stats.

```python
df = pd.read_csv(r'../input/world-happiness-report-2021/world-happiness-report-2021.csv')
df.head()
```

```python
world_map_js = r'../input/d/sivasankarmc/world-happiness-report-2021/custom.geo.json'

print(type(world_map_js))
```

* ## Datalayer addition with Choropleth
* ### choropleth method under folium to add data visual layer
* ####  specify the .geojson file loaded var in *geo_data* *arg*
* ####  specify the dataset imported containing the stats to be displayed in *data* *arg*
* ####  *columns* *arg* to specify the columns to be displayed in the dataset,
* ####  *key_on* *arg* to specify the .geojson file feature under which the columns data to be displayed - in this case it is properties.name
* ####  *other* *arg*s are fill_color, fill_opacity, legend_name, smooth_factor
* ####  with *.add_to* method we can add this layer to the folium.map.

* #### In below example will display the world gdp data stats distribution across the world above the Folium worldmap layer

```python
folium.Choropleth(geo_data=world_map_js ,data=df, columns=['Country name','Logged GDP per capita'],key_on='feature.properties.name',fill_color = 'RdPu',fill_opacity=1,line_opacity=1,legend_name='World GDP per capita - Log',smooth_factor=0).add_to(Worldmap)

Worldmap
```
