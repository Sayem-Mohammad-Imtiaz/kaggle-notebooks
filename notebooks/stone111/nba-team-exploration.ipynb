## 1. Package and Data Loading

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
from scipy.stats import norm
color = sns.color_palette()
from IPython.core.display import display, HTML
from ggplot import *
display(HTML("<style>.container { width:100% !important; }</style>"))
%matplotlib inline
```

```python
players_with_salary_wiki_twitter_df = pd.read_csv("../input/nba_2017_players_with_salary_wiki_twitter.csv", index_col = 0)
players_with_salary_wiki_twitter_df.head()
```

## 2. Correlation Checking with Heatmap

```python
plt.subplots(figsize=(10,10))
ax = plt.axes()
ax.set_title("NBA Player Correlation Heatmap: Top 20 Metrics")
corr = df.corr()
columns = corr.nlargest(20, 'WINS_RPM')['WINS_RPM'].index
coff = np.corrcoef(df[columns].values.T) 
sns.heatmap(coff, cbar=True, annot=True, square=True, fmt='.2f', 
                      annot_kws={'size': 8},  
                      yticklabels=columns.values, xticklabels=columns.values)
```

**From the heatmap above, we can see that the salary predict winning metrics better than social media. Therefore, we build the model based on salary**

## 3. Linear Modeling Building

```python
sns.lmplot(x='SALARY_MILLIONS', y='WINS_RPM', data = df)
```

```python
sns.distplot(df['SALARY_MILLIONS'])
```

```python
#Normalizing by log
df['SALARY_MILLIONS_Log'] = np.log(df['SALARY_MILLIONS'])
sns.distplot(df['SALARY_MILLIONS_Log'], fit = norm)
```

```python
lm = smf.ols(formula='WINS_RPM ~ SALARY_MILLIONS_Log', data = df).fit()
print(lm.summary())
```

** Even though the salary predict winning metrics better than social media, the linear model does fit well with a lower R-square value. I think the Social Media is a good tool to analyze the winning metrics since it is more interactive with their performance.
