```python
# Showing the numerical varibales with the highest correlation with 'budget', sorted from highest to lowest
correlation = data.select_dtypes(include=[np.number]).corr()
print(correlation['budget'].sort_values(ascending = False))
```

```python
# Heatmap of correlation of numeric features
f , ax = plt.subplots(figsize = (14,12))
sns.heatmap(data.corr(), annot=True, square = True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```

```python
# Pair plot
sns.set()
cols = ['budget', 'id', 'popularity', 'revenue', 'runtime', 'vote_average', 'vote_count']
sns.pairplot(data[cols],size = 2 ,kind ='scatter',diag_kind='kde')
plt.show()
```

```python
# Line Plot
# color = color, label = label, linewidth = width of line, alpha = opacity, grid = grid, linestyle = sytle of line
data.budget.plot( kind = 'line', color = 'g',label = 'budget',linewidth=1,alpha = 0.5,grid = True,linestyle = '-')
data.runtime.plot( color = 'r',label = 'runtime',linewidth=1, alpha = 0.5,grid = True,linestyle = 'steps')
```

```python
# Scatter Plot 
# x = budget, y = popularity
data.plot(kind='scatter', x='budget', y='popularity',alpha = 0.5,color = 'red')
plt.xlabel('budget')              # label = name of label
plt.ylabel('popularity')
plt.title('Budget Popularity Scatter Plot')            # title = title of plot
```

```python
# Histogram
# bins = number of bar in figure
data.id.plot(kind = 'hist',bins = 50,figsize = (12,12))
plt.show()
```
