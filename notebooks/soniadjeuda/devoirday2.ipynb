```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
Data_filepath = "../input/hr-analytics-job-change-of-data-scientists/aug_train.csv"

# Read the file into a variable flight_data
myDAta= pd.read_csv(Data_filepath)
myDAta.head()
```

```python

# titre du graph
plt.title("relevent_experience Vs myData.training_hours")

# Bar chart par gendre
sns.barplot(x=myDAta.relevent_experience, y=myDAta.training_hours, hue="target", data=myDAta)
```

```python
# titre du graph
plt.title("myData.relevent_experience Vs myData.training_hours")

# Bar chart par gendre
sns.barplot(x=myDAta.relevent_experience, y=myDAta.training_hours, hue="target", data=myDAta)
```

```python
plt.title("myData.education_level Vs myData.training_hours")

sns.barplot(x=myDAta.target, y=myDAta.training_hours, hue="education_level", data=myDAta)
```
