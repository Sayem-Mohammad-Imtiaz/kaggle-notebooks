###**Importing Required Libraries**

```python
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

###**Loading and Analyzing the dataset**

```python
df = pd.read_csv('who_suicide_statistics.csv') #read_csv function is used to read the dataset into a pandas dataframe

df.head() #Prints the first 5 rows of the data
```

```python
print('Number Of Rows:',df.shape[0])
print('Number Of Columns:', df.shape[1])
```

```python
df.nunique()
```

```python
a = df.columns
for i in range(0, len(a)):
  print('Name of Column',i+1,':',a[i])
```

```python
df.isnull().sum()
```

```python
df = df.dropna() #Dropping Rows if the row contains all NaN Values

df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

###**Data Visualization**

###**Plotting Univariate Distribution**

- Histogram
- Countplot

####**Histogram**

#####Using **Seaborn**

```python
sns.set()
plt.figure(figsize=(12,8))
sns.distplot(df['year'], kde = False)

plt.xlabel('Year')
plt.ylabel('Count')

plt.show() 
```

#####Using **Plotly Express**

```python
fig = px.histogram(df, x="year")
fig.show()
```

####**Countplot**

```python
plt.figure(figsize = (12,8))
ax = sns.countplot(df["year"], color='green')

plt.xticks( rotation=90 )

plt.show()
```

###**Plotting Bi-variate Distribution**

A Bi-variate distribution is a distribution of two random variables

The concept generalizes to any number of random variables, giving a **Multivariate Distribution**

- Bar Plot
- Scatter Plot
