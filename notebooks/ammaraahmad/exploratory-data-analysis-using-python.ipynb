```python
#In this notebook,We learn about Exploratory Data Analysis(EDA) and Steps to complete this process.
#Firstly we understand why Exploratory Data Analysis is important.EDA helps us in understanding and cleaning data.
#EDA also helps us in analysing relationship between different variables.So,lets get started.
```

```python
#Importing Liabraries
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt  
```

```python
#Importing dataset from kaggle which is "StudentsPerformance.csv"
data=pd.read_csv("../input/students-performance-in-exams/StudentsPerformance.csv")
```

```python
#1.Understanding the Data
#First step in EDA is understanding the data
data.head()
```

```python
data.tail()
```

```python
#Here you can see that dataset consists of 1000 rows and 8 columns
data.shape
```

```python
#Basic detail about dataset
data.describe()
```

```python
#It shows total columns in dataset 
data.columns
```

```python
#Check unique values in all columns
data.nunique()
```

```python
#Check unique values in specific column
data['parental level of education'].unique()
```

```python
#2.Cleaning the Data
#Second step in EDA is to clean Data
```

```python
#Check if any column has null value
data.isnull().sum()
```

```python
#Droping unnecessary Data
student= data.drop(['race/ethnicity','parental level of education'],axis=1)
student.head()
```

```python
#3.Relationship Analysis
#In this step,we try to find relationship between variables 
```

```python
sns.pairplot(student)
```

```python
#This will show relation between two columns with respect to gender
sns.relplot(x='math score',y='reading score',hue='gender',data=student)
```

```python
sns.distplot(student['math score'],bins=5)
```

```python
student=pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
student.head()
sns.scatterplot(x="reading score",y="math score",data=student,hue="writing score")
plt.show()
```

```python
student=pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
student.head()
sns.scatterplot(x="reading score",y="math score",data=student,hue="lunch",style="lunch")
plt.show()
```

```python
sns.set(style="whitegrid")
student=pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
sns.barplot(x="test preparation course",y="math score",data=student,color='green')
plt.show()
```

```python
sns.lineplot(x="math score",y="writing score",data=student,hue='gender',style="gender",markers=True)
plt.show()
```

```python
plt.plot('math score','reading score',data=student,color='r',linestyle=':',linewidth=2)
plt.plot('math score','writing score',data=student,color='g',linestyle=':',linewidth=2)
plt.title("Two lines in graph")
plt.ylabel("y-axis")
plt.xlabel("x-axis")
plt.grid(True)
plt.show()
```
