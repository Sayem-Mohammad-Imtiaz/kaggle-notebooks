## Search Strategy:
In general, the step-by-step method for writing a literature review includes: 
1. Deciding on a focus area and question
2. Choosing a database to search 
3. Narrow topic and select papers by finding relevant excerpts/study designs in articles 
4. Organize selected articles to look for patterns of evidence in articles by developing subtopics
5. Evaluate papers
6. Write

Our search engine accelerates the time and effort needed to conduct literature reviews by skipping steps 2 - step 4. After developing a research question, the search engine automates each step to allow researchers to quickly begin analyzing articles and begin their own rapid hypothesis testing.

### **Articulating complex information needs using query templates**

When researchers use search engines, many questions can go unanswered due to lack of skills in formulating questions, crafting effective search strategies, and identifying the best levels of evidence. Formulating a well-focused question is the first and arguably most important step.  This search engine was designed to help researchers develop well-focused questions from the broad tasks given in the [COVID-19 Open Research Dataset Challenge](https://www.kaggle.com/allen-institute-for-ai/CORD-19-research-challenge/tasks) through evidence-based practice. 

**Framing tasks into research engine questions**: The PICOT framework is known to help researchers retrieve more precise, relevant results by formatting a question into 5 components:![](http://i.imgur.com/9cEw6YG.png)

It allows better specificity and conceptual clarity to the tasks given in the challenge because it dissects the questions into smaller components, which are then easier to search. As we aim to solve complex problems, our search engine provides guidance by *nudging* its' users to ask the right questions.

### Information Retrieval: The Great Search for Knowledge

The aim of information retrieval (IR) is to make machine-stored data discoverable by filtering specific information from a set of data. While there are various information retrieval models, this engine uses a vector space model where documents and search queries are represented as vectors or matrices. The documents and search queries receive values related to the search term and can be represented as points within the vector space. As vectors are compared to each other, the vector with the content closest to that of the search query will appear first in the result rankings. The advantage of this approach is speed and simplicity: once a document index is built, terms can be searched quickly, and there is no need to train a model or annotate complex datasets. However, the disadvantage is that without boolean operators, no terms can be excluded.

**The gains:** BM25 is a ranking function for information gain used by search engines to rank matching documents according to their relevance to a given search query. It is one of the strongest “simple” scoring functions where BM25 scores combine the query term frequency and the document term frequency, normalized by the ratio of the document’s length, to the average length.

### Evidence-based AI: Automated screening for literature using evidence-based practice query

Once a research question as been formulated, the search engine provides the most relevant articles with automated screening of PICOT characteristics. The engine will query each PICOT term individually at sentence level to ensure that when searching for multi-words they all appear in the same sentence, then compute the intersection at paper level (all the papers that had at least one sentence found in each PICOT query).
![](http://i.imgur.com/BdV4Bsb.png)

**Screening Evaluation:** We experimented with different **(P)(I)(C)(O)(T)** combinations, which applied different thresholds for the number of study characteristics needed to be included. In our research, we found that:

- All **(P)(I)(C)(O)(T**) terms performed the worst with very few matches
- Screening with any 3 **(P)(I)(C)(O)** terms improved recall
- Screening for 2 terms provided the best precision for relevant search results
  * **(P)(I)** performed best with the most relevant articles
  * **(I)(O)** was more precise than using any 2 **(P)(I)(C)(O)**
- Using **(T)** limited the results as it had the least number of sentence matches

<font size="2" color="black"> This inital evaluation can be used as guiding principles to using the engine since our tool allows researchers to choose their own inclusion/exclusion criteria for the terms in the final result. We ultimately measured accuracy in value of retrieving actionable results for researchers and contributing to the Kaggle Community.</font>

## 2. Initialization

Import required libraries and set required parameters, namely:
* max_docs: the maximum number of documents (sentences) to return per query
* threshold: the minimum score a document must be given to be returned (note a 0 score is given to completely irrelevant sentences)
* paper_folders: path to each folder having a set of CORD19 papers in json format
* sentence_folders: path to each folder where to save the corresponding json files with the sentences extracted per paper
* sentence_file: path to file where to save the dataframe containing all the sentences in all the papers, including the paper they come from
* index_file: path file where to save the BM25 index of all the sentences in the dataframe
* metadata_file: path to CORD19 metadata file

```python
import os
import sys
import pandas as pd

from ask_jc.paper.paper_sentence_extractor import PaperSentenceExtractor
from ask_jc.paper.paper_sentence_dataframe import PaperSentenceDataFrame
from ask_jc.indexer.bm25_indexer import Bm25Indexer
from ask_jc.searcher.bm25_searcher import Bm25Searcher
```

```python
pd.set_option('display.max_colwidth', 0) # Do not truncate contents of tables when printing them

recompute_index = False # do not recompute index if already present
max_docs = 0 # max_docs 0 means there is no limit on the maximum amount of documents to return per query
threshold = 0.1 # keep documents that have some score at least slightly above 0
# datasets = ['biorxiv_medrxiv', 'comm_use_subset', 'custom_license', 'noncomm_use_subset'] # Kaggle runs out of RAM with all datasets
datasets = ['biorxiv_medrxiv', 'custom_license'] # As an example, we used here these 2 datasets only so the notebook can run in Kaggle
cord19_folder = os.path.join('..', 'input', 'CORD-19-research-challenge')
cord19index_folder = os.path.join('..', 'input', 'cord19index')
output_folder = os.path.join('..', 'output')
paper_folders = [os.path.join(cord19_folder, folder, folder, 'pdf_json') for folder in datasets]
sentences_folder = os.path.join(output_folder, 'sentences')
sentence_folders = [os.path.join(sentences_folder, folder) for folder in datasets]
sentence_file = os.path.join(cord19index_folder, 'sentences.gzip') # use precomputed sentences dataframe if exists
if recompute_index or not os.path.exists(sentence_file):
    sentence_file = os.path.join(output_folder, 'sentences.gzip') # otherwise generate it in output folder
index_file = os.path.join(cord19index_folder, 'bm25_index.pkl') # use precomputed BM25 index if exists
if recompute_index or not os.path.exists(index_file):
    index_file = os.path.join(output_folder, 'bm25_index.pkl') # otherwise generate it in output folder
metadata_file = os.path.join(cord19_folder, 'metadata.csv')
metadata = pd.read_csv(metadata_file)
```

## 3. Execute Sentence Extractor
**Sentence Splitting:** To be able to retrieve the results at the sentence level, we need to pre-process the data to define and extract the sentences from the papers. We associated these sentences with the metadata to help explore the search engine results in step 7. (We also found it useful to help correspond the section and paper with references later on).

```python
if recompute_index or not os.path.exists(sentence_file):
    extractor = PaperSentenceExtractor()
    for p, s in zip(paper_folders, sentence_folders):
        extractor.extract_and_save(p, s)
    del(extractor)
```

## 4. Make dataframe
Load sentences into pandas dataframe and save file as compressed pkl

```python
if recompute_index or not os.path.exists(sentence_file):
    sentence_df = PaperSentenceDataFrame()
    sentence_df.load_folder(sentences_folder)
    sentence_df.save_pickle(sentence_file)
    del(sentence_df)
```

## 5. Execute Indexer
The indexer loads the previously saved pandas dataframe and creates a BM25 index for all the sentences and saves it to a file. Note this step takes some time to run and save the index file, please be patient. Once the index is saved to a file, one can re-run the searcher without having to recompute the index as long as the set of papers is not changed.

```python
if recompute_index or not os.path.exists(index_file):
    indexer = Bm25Indexer()
    indexer.index_dataframe(sentence_file)
    indexer.save_index(index_file)
    del(indexer) # Delete indexer once the index is built, since we will use the searcher from now on on the saved index
```

## 6. Search
First of all the sentences dataframe and the index file have to be loaded, this takes some time so please be patient. Afterwards queries can be performed, which will be returned as a subset of the sentences dataframe with an additional score column. The returned dataframe can then be joined with the metadata file to retrieve the information of the paper the found sentences belong to.

```python
searcher = Bm25Searcher()
searcher.load_sentences_and_index(sentence_file, index_file)
```

```python
hits = searcher.search('example query', max_docs=max_docs, threshold=0.1)
hits
```

## 7. Explore
We use the **searcher** here to perform each one of the (P)(I)(C)(O)(T) queries at sentence level, then compute the intersection of the papers of each individual query to find the papers that have at least one sentence matching some or all of the (P)(I)(C)(O)(T) queries. There are 5 separate lines of code to run the searcher 5 times for each element. After exploring the results of each, the researcher can decide how many and which PICOT terms to use in the final search.

```python
#note that the number of documents found will appear at the bottom to help you customize which terms you'd like to include in final search
#POPULATION
hitsP = searcher.search('covid-19 coronavirus', max_docs=max_docs, threshold=threshold)
hitsP
```

```python
#INTERVENTION
hitsI = searcher.search('molecular based assays serological tests RNA-based antibodies', max_docs=max_docs, threshold=threshold)
hitsI
```

```python
#COMPARISON
hitsC = searcher.search('reverse transcription-polymerase chain reaction (RT-PCR) immunoassay chest xray', max_docs=max_docs, threshold=threshold)
hitsC
```

```python
#OUTCOME
hitsO = searcher.search('rapid point-of-care detection diagnosis', max_docs=max_docs, threshold=threshold)
hitsO
```

```python
#TYPE OF STUDY 
hitsT = searcher.search('comparison', max_docs=max_docs, threshold=threshold)
hitsT
```

```python
def get_common_paper_results(hits_list, metadata):
    unique_ids_list = [hits[['paper_id']].drop_duplicates() for hits in hits_list]
    common_paper_ids = pd.concat(unique_ids_list, axis=1, join='inner').iloc[:, 0].to_frame()
    common_hits_list = [hits.loc[hits['paper_id'].isin(common_paper_ids['paper_id'])] for hits in hits_list]
    common_hits_list = [common_hits.merge(metadata, how='left', left_on='paper_id', right_on='sha') for common_hits in common_hits_list]
    common_hits_list = [common_hits[['doi', 'title','authors', 'journal','publish_time','abstract', 'paper_id', 'sentence']] for common_hits in common_hits_list]
    return common_hits_list
```

```python
#identify which PICOT terms to use for final search, this only uses (P)atient, (I)ntervention,(C)omparison
final_search = get_common_paper_results([hitsP,hitsI,hitsC], metadata)
```

```python
#use PICOT terms passed in final_search to pass to original queries to explore the sentences used to link the common articles
resultsP, resultsI, resultsC = get_common_paper_results([hitsP, hitsI, hitsC], metadata)
```

# Conclusion
We demonstrated a novel way to automate the workflow of conducting scientific literature review with an AI assisted search engine. This engine takes into consideration the everyday challenges of conducting reviews and prioritizes results uniquely to each user through an evidence-based framework to provide *a better research experience*. We believe that the combination of nudging users to reframe their research goals and the processing speed of BM25 will save time and resources needed to review articles during this crisis. While a main limitation to this approach was the susceptibility for type I errors (false positives/irrelevant studies), future development will examine the use of a grammar engine and more advanced AI to avoid such errors and test the effectiveness of other methods.
