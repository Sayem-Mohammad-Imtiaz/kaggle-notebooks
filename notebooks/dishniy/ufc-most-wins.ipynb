```python
data = pd.read_csv('/kaggle/input/ufcdata/raw_total_fight_data.csv', sep=';')
data.head()
```

```python
winner = data[['Winner']]
winner = winner.dropna()
winner
```

```python
win_dict = {}

for i in winner.Winner:
    if i not in win_dict:
        win_dict.update({i : 1})
    else:
        win_dict[i] +=1
```

```python
sorted_d = sorted(win_dict.items(), key=lambda x: x[1], reverse= True)
sorted_d
```
