```python
# as you can see there is the column named 'course_students_enrolled' 
# having k with each value lets change this with 1000 so, that column look normal
coursera['course_students_enrolled']=coursera['course_students_enrolled'].replace({'k': '*1e3', 'm': '*1e6'}, regex=True).map(pd.eval).astype(int)
# print first five rows of column 'course_students_enrolled'
coursera['course_students_enrolled'][:5]
```

```python
# rename the column
coursera=coursera.rename(columns={'Unnamed: 0':'unknown'})
# as unknown column is not needed so drop this column
coursera=coursera.drop(['unknown'],axis=1)
coursera.head()
```

### Exploring University of Michigan (I have taken the courses on coursera)

```python
# take a look on University of Michigan for example checking the how many courses from this university etc.

UOM=coursera[coursera['course_organization']=='University of Michigan']
# print(UOM)
# how many courses and specialization by university of michigan
len(UOM)
```

```python
# now take a look on course certificate type of university of michigan
UOM['course_Certificate_type'].value_counts()
```

```python
# lets check total strenght of students enrolled in all courses of university of michigan
UOM['course_students_enrolled'].sum()
```

```python
# so, lets check the top 10 courses which have greater enrollment of students in university of michigan.
course_enrolled=UOM[['course_title','course_students_enrolled']]
# Now, time to check the top ten courses or specialization which have greater students enrolled
top_enrolled=course_enrolled.nlargest(10,['course_students_enrolled'])
top_enrolled
# bar plot for these courses
x=top_enrolled['course_title']
y=top_enrolled['course_students_enrolled']
plt.figure(figsize=(8,5))
bar_color=['#D2691E','FireBrick','#4682B4','#778899','#66CDAA','cyan','pink','#90EE90','MediumPurple','Plum']
plt.barh(x,y,color=bar_color)
plt.gca().invert_yaxis()
plt.xlabel('student enrolled')
plt.title('Most students enrolled in the courses of university of michigan')
plt.show()
```

```python
# lets check the course difficulty of courses of university of michigan
UOM['course_difficulty'].value_counts()
```

```python
# lets check course_difficulty of the courses and specialization of university of michigan or 
# check which course on which course_difficulty

m=pd.DataFrame(UOM.groupby(['course_difficulty','course_title']).groups)
m.T
```

### Now again Exploring according to whole dataset

```python
# checking the top four organization along with particular course_title who have higher number of student enrolled
organization_enrolled=coursera[['course_organization','course_title','course_students_enrolled']]
top_enrolled=organization_enrolled.nlargest(5,['course_students_enrolled'])
top_enrolled

# plotting
organization=top_enrolled['course_organization']
enrolled=top_enrolled['course_students_enrolled']
plt.figure(figsize=(9,3))
sns.barplot(x=organization,y=enrolled)
plt.show()
```

```python
# as we can see the stanford university on the top of who getting maximum student enrolled in one particulr course

# lets check the which university organization have how many courses on coursera
organization=coursera['course_organization'].value_counts().sort_values(ascending=False)
organization=organization.head(5)
organization=organization.reset_index()
organization.columns=['course_organization','number of courses']

plt.barh(organization['course_organization'],organization['number of courses'])
plt.gca().invert_yaxis()
```

```python
# now its time to check the top 5 organization who have maximum student enrollment in all the courses 
# they have on the coursera
total_enrollment=pd.DataFrame(coursera.groupby(['course_organization'])['course_students_enrolled'].sum().sort_values(ascending=False))
total_enrollment=total_enrollment[:5]
print(total_enrollment)

# plotting
x=total_enrollment.index
y=total_enrollment['course_students_enrolled']
plt.figure(figsize=(8,6))
plt.barh(x,y)
plt.gca().invert_yaxis()
plt.title('top organizations which have maximum enrollments in all their courses on coursera')
plt.show()
```
