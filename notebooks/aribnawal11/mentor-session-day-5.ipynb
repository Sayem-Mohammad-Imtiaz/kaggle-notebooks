```python
#convert the DATE_TIME column from string datatype(OBJECT) to Datetime Datatype
df['DATE_TIME']= pd.to_datetime(df['DATE_TIME'],format = '%d-%m-%Y %H:%M')
df.info()
```

```python
#Split the DATE_TIME column into Date column separately
df['DATE']= pd.to_datetime(df['DATE_TIME'],format = '%d-%m-%Y %H:%M').dt.date
df.info()
```

```python
#Split the DATE_TIME column into Time column separately
df['TIME']= pd.to_datetime(df['DATE_TIME'],format = '%d-%m-%Y %H:%M').dt.time
df.info()
```
