### 1. Let us see top viloation locations for the current year
``
 1.1 Make violation date as index
 1.2 Index by current year
 1.3 Group violations by intersection for currrent year
 1.4 Plot the data
``

```python
# Violation date is string so we need to change it to datetime to make use of python's date time indexing features
red_light_vio['VIOLATION DATE'] = pd.to_datetime(red_light_vio['VIOLATION DATE'])
red_light_vio.sort_values(by='VIOLATION DATE',inplace=True)
red_light_vio.set_index('VIOLATION DATE',inplace=True)
red_light_vio.head()
```

```python
curr_yr_tot = red_light_vio.loc['{}'.format(pd.datetime.now().year)].groupby(['INTERSECTION']).VIOLATIONS.sum()
curr_yr_tot.head()
```

```python
plt.grid(True)
ax = curr_yr_tot.sort_values(ascending=False)[0:10].plot(kind='bar')
ax.grid(True,axis='y')
plt.title('Top ten redlight violation locations in the current year')
plt.show()
```

### 2. Let's see total violations per year

```python
 red_light_vio.groupby(lambda x: x.year).VIOLATIONS.sum().plot(kind='bar')
```
