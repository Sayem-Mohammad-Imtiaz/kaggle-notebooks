```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import keras
from keras.models import Sequential
from keras.layers import Dense, Flatten

%matplotlib inline
```

```python
df_train = pd.read_csv("../input/mnist-in-csv/mnist_train.csv")
df_test = pd.read_csv("../input/mnist-in-csv/mnist_test.csv")
```

```python
df_train
```

```python
X_t = df_train.drop(["label"], axis=1)
Y_t = df_train.label.values
```

```python
X_test = df_test.drop(["label"], axis=1)
Y_test = df_test.label.values
```

```python
plt.figure(figsize=(6,3))
X = X_t.to_numpy().reshape(60000, 28, 28)
for i in range(8):
    plt.subplot(2,4,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X[i], cmap=plt.cm.binary)
    plt.xlabel(str(Y_t[i]))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_val, Y_train, Y_val = train_test_split(X_t, Y_t, test_size=0.2, stratify = Y_t, random_state = 42)
```

```python
num_classes = 10
X_train = X_train/255.
X_val = X_val/255.
X_test = X_test/255.
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_val = keras.utils.to_categorical(Y_val, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)
```

```python
Y_val
```

```python
col = 28
row = 28
input_shape = (row, col, 1)
```

```python
X_train = X_train.to_numpy().reshape(X_train.shape[0], col, row, 1)
X_val = X_val.to_numpy().reshape(X_val.shape[0], col, row, 1)
X_test = X_test.to_numpy().reshape(X_test.shape[0], col, row, 1)
```

```python
X_test.shape
```

```python
model = Sequential([
    Flatten(input_shape=(28,28,1)),
    Dense(1024, activation='relu'),
    Dense(10, activation='softmax'),
])

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.Adadelta(), metrics=['accuracy'])
```

```python
X_train.shape
Y_train.shape
```

```python
model.fit(X_train, Y_train,
          batch_size=32,
          epochs=20,
          verbose=1, 
          validation_data=(X_val, Y_val))
```

```python
score = model.evaluate(X_test, Y_test, verbose=0)
print('Validation loss:', score[0])
print('Validation accuracy:', score[1])
```
