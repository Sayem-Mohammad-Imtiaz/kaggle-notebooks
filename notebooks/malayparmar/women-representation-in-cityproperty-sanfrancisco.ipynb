```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from collections import Counter
plt.style.use("fivethirtyeight")
output = pd.DataFrame(columns=['Facility', 'No. of Males', 'No. of Females'])
data = pd.read_csv('/kaggle/input/women-representation-in-city-property-sanfrancisco/WomenRepresentaionInCityProperty-SanFrancisco.csv')
mod = data[['Department/Source', 'Gender']]
facility = data['Department/Source'].unique()
mod = mod.groupby(by='Department/Source')
for item in facility:
    gen_counter = Counter()
    for x in mod.get_group(item)['Gender']:
        gen_counter.update(x.split())
    output = output.append({'Facility': item, 'No. of Males': str(gen_counter['M']), 'No. of Females': str(gen_counter['F'])}, ignore_index=True)
x_indexes = np.arange(len(list(output['Facility'])))
width = 0.4
count_m = list(output['No. of Males'])
count_f = list(output['No. of Females'])
for i in range(0, len(count_m)):
    count_m[i] = int(count_m[i])
    count_f[i] = int(count_f[i])
plt.bar(x_indexes - (width/2), count_m, width=width, color="#008fd5", label="No. of Males")
plt.bar(x_indexes + (width/2), count_f, width=width, color="#e5ae38", label="No. of Females")
plt.legend()
plt.xticks(ticks=x_indexes, labels=list(output['Facility']), rotation=45)
plt.tick_params(labelsize=10)
plt.title("No. of Names of Males and Females in Different Facilities")
plt.xlabel("Facilities", fontsize=14)
plt.ylabel("No. of Males/Females", fontsize=14)

plt.tight_layout()
plt.show()
print(output)
print("Total no. of Male Names:",sum(count_m))
print("Total no. of Female Names:",sum(count_f))
```

As visible from the output, there are total 64 male names and 28 female names in the Property Names of San Francisco.
It is observed that in each facility the no. of male names is greater than or equal to no. of female names.
The highest number of female names is present in 'Administrator' facility.
There are 3 facilities namely 'AIRPORT', 'PUC' and 'SFMTA' with zero female names present in them.
