```python
new_data=stroke_data.drop(columns=['bmi','id'])
```

```python
new_data
```

```python
classification_setup = classification.setup(data= new_data, target='stroke')
```

```python
classification_data = classification.create_model('dt')
```

```python
classification_xgb = classification.create_model('xgboost')
```

```python
classification.compare_models()
```

```python
lr= classification.create_model('lr')
```

```python
tuned_lr = classification.tune_model(lr)
```

```python
classification.tune_model(lr, optimize = 'AUC') 
```

```python
classification.plot_model(tuned_lr, plot = 'auc')
```

```python
classification.plot_model(tuned_lr, plot='feature')
```

```python
classification.evaluate_model(tuned_lr)
```

```python
classification.predict_model(tuned_lr);
```

```python
predictions = classification.predict_model(tuned_lr, data=stroke_data)
```

```python
predictions
```

```python
final_model = classification.finalize_model(tuned_lr)
```

```python
classification.predict_model(final_model)
```

```python
classification.save_model(final_model,'Stroke prediction')
```
