```python
def get_income(Latitude,Longitude):
    pixel_y=(Latitude-40.915308)/(40.494236-40.915308)*image.shape[0]
    pixel_x=(Longitude-(-74.265914))/(-73.690971-(-74.265914))*image.shape[1]
    colour=image[int(pixel_y)][int(pixel_x)]
    r=colour[0]*250
    g=colour[1]*250
    b=colour[2]*250
    minimum=abs(247-r)+abs(251-g)+abs(254-b)
    income=15000
    
    if(abs(226-r)+abs(238-g)+abs(250-b)<minimum):
        minimum=abs(226-r)+abs(238-g)+abs(250-b)
        income=25000
    if(abs(205-r)+abs(221-g)+abs(238-b)<minimum):
        minimum=abs(205-r)+abs(221-g)+abs(238-b)
        income=35000
    if(abs(175-r)+abs(209-g)+abs(231-b)<minimum):
        minimum=abs(175-r)+abs(209-g)+abs(231-b)
        income=45000
    if(abs(136-r)+abs(190-g)+abs(220-b)<minimum):
        minimum=abs(136-r)+abs(190-g)+abs(220-b)
        income=55000
    if(abs(95-r)+abs(166-g)+abs(209-b)<minimum):
        minimum=abs(95-r)+abs(166-g)+abs(209-b)
        income=65000
    if(abs(61-r)+abs(141-g)+abs(195-b)<minimum):
        minimum=abs(61-r)+abs(141-g)+abs(195-b)
        income=75000
    if(abs(33-r)+abs(113-g)+abs(181-b)<minimum):
        minimum=abs(33-r)+abs(113-g)+abs(181-b)
        income=85000
    if(abs(15-r)+abs(82-g)+abs(150-b)<minimum):
        minimum=abs(15-r)+abs(82-g)+abs(150-b)
        income=95000
    if(abs(8-r)+abs(48-g)+abs(107-b)<minimum):
        minimum=abs(8-r)+abs(48-g)+abs(107-b)
        income=105000
    return income
```

```python
satscore=satscore.dropna()
satscore["income"]=""
satscore["Start Time F"]=""
satscore["End Time F"]=""
satscore["Hours"]=""
for i in range(len(satscore)):
    satscore.iloc[i,satscore.columns.get_loc("income")]=get_income(satscore.iloc[i]['Latitude'],satscore.iloc[i]['Longitude'])
    
    temp=satscore.iloc[i]["Start Time"].split(" ")
    temp1=temp[0].split(":")
    timeS=int(temp1[0])*60+int(temp1[1])
    if(temp[1]=='PM'):
        timeS=timeS+12*60
    satscore.iloc[i,satscore.columns.get_loc("Start Time F")]=timeS
    
    temp=satscore.iloc[i]["End Time"].split(" ")
    temp1=temp[0].split(":")
    timeE=int(temp1[0])*60+int(temp1[1])
    if(temp[1]=='PM'):
        timeE=timeE+12*60
    satscore.iloc[i,satscore.columns.get_loc("End Time F")]=timeE
    
    satscore.iloc[i,satscore.columns.get_loc("Hours")]= timeE-timeS
    
    satscore.iloc[i,satscore.columns.get_loc("Percent White")]=satscore.iloc[i]["Percent White"][:-1]
    satscore.iloc[i,satscore.columns.get_loc("Percent Black")]=satscore.iloc[i]["Percent Black"][:-1]
    satscore.iloc[i,satscore.columns.get_loc("Percent Hispanic")]=satscore.iloc[i]["Percent Hispanic"][:-1]
    satscore.iloc[i,satscore.columns.get_loc("Percent Asian")]=satscore.iloc[i]["Percent Asian"][:-1]
    
    satscore.iloc[i,satscore.columns.get_loc('Student Enrollment')]=float(satscore.iloc[i]["Percent Tested"][:-1])/100*satscore.iloc[i]['Student Enrollment']
satscore.head()
```

```python
from sklearn.model_selection import train_test_split
X = satscore[X_features]
y = satscore[Y_features[0]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
# Create model
linreg = LinearRegression()
linreg.fit(X_train, y_train)
```

```python
y_hat_test = linreg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
# See our Squared Mean Error score and Root Mean Squared Error:
test_mse = mean_squared_error(y_test, y_hat_test)
test_rmse = np.sqrt(test_mse)
# See our Mean Absolute Error
test_mae = mean_absolute_error(y_test, y_hat_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
```

```python
sregr=[]
score=[]
for depth in range(1,30):
    regr = RandomForestRegressor(n_estimators=depth, random_state=0)
    regr.fit(X, y)
    score.append(regr.score(X_test, y_test, sample_weight=None))
    sregr.append(regr)
```

```python
for depth in range(1,30):
    print("Dept:",depth," Acuracy=",score[depth-1]*100,"%")
```
