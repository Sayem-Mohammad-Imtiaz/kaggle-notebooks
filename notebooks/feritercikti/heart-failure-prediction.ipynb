```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
dataset = pd.read_csv("../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv")
```

We need to delete the 'time' column, as this feature will not help us for prediction.

```python
dataset.drop('time',axis=1)
```

# Training and Testing 


```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
X_train, X_test, y_train , y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

**Feature Scaling**

```python
sc = MinMaxScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

#  Logistic Regression

```python
logclassifier = LogisticRegression(random_state=0)
logclassifier.fit(X_train,y_train)
```

```python
y_pred = logclassifier.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

# K-Nearest Neighbors


```python
knnclassifier = KNeighborsClassifier(n_neighbors = 5,metric = 'minkowski', p =2)
knnclassifier.fit(X_train,y_train)
```

```python
y_pred = knnclassifier.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

# Support Vector Machine

```python
svmclassifier = SVC(kernel='linear', random_state= 0)
svmclassifier.fit(X_train,y_train)
```

```python
y_pred = svmclassifier.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

# Kernel SVM

```python
ksvmclassifier = SVC(kernel='rbf', random_state= 0)
ksvmclassifier.fit(X_train,y_train)
```

```python
y_pred = ksvmclassifier.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

# Naive Bayes

```python
nbclassifier = GaussianNB()
nbclassifier.fit(X_train,y_train)
```

```python
y_pred = nbclassifier.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

# Decision Tree Classification

```python
dtclassifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
dtclassifier.fit(X_train, y_train)
```

```python
y_pred = dtclassifier.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```

# Random Forest Classification

```python
rfclassifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
rfclassifier.fit(X_train, y_train)
```

```python
y_pred = rfclassifier.predict(X_test)

cm = confusion_matrix(y_test,y_pred)
print(cm)
accuracy_score(y_test,y_pred)
```
