```python
# 1 - Filtering Pandas data frame
x = data['Global_Sales']>40     # There are only 2 pokemons who have higher global_sales value than 40
data[x]
```

```python
# 2 - Filtering pandas with logical_and

data[np.logical_and(data['Global_Sales']>25, data['Year']>2000.0 )]
```

```python
#Same code

data[(data['Global_Sales']>25) & (data['Year']>2000.0)]
```

```python
#NA_Sales each i if > NA_Sales.mean() we create a new column and write  high
# or low
threshold = sum(data['NA_Sales'])/len(data['NA_Sales'])
data["sales_level"] = ["high" if i > threshold else "low" for i in data['NA_Sales']]
data.loc[::-1,["sales_level","NA_Sales"]] 
```
