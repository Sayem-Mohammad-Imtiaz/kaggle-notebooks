```python
df=df.drop(['YEAR','FLIGHT_NUMBER','AIRLINE','DISTANCE','TAIL_NUMBER','TAXI_OUT','SCHEDULED_TIME','DEPARTURE_TIME',
            'WHEELS_OFF','ELAPSED_TIME','AIR_TIME','WHEELS_ON','DAY_OF_WEEK','TAXI_IN','ARRIVAL_TIME',
            'ORIGIN_AIRPORT','CANCELLATION_REASON'],axis=1)
```

```python
df.isnull().sum()[df.isnull().sum()!=0]
```

```python
df=df.fillna(df.mean())
```

```python
df[df['ARRIVAL_DELAY']>15]
```

```python
d=[]
for i in df['ARRIVAL_DELAY']:
    if i > 15:
        d.append(1)
    else:
        d.append(0)
```

```python
df['result'] = d
```

```python
df['result'].value_counts()
```

```python
df.info()
```

```python
x=df[['MONTH','DAY','SCHEDULED_DEPARTURE','DEPARTURE_DELAY','SCHEDULED_ARRIVAL','DIVERTED','CANCELLED','AIR_SYSTEM_DELAY',
    'SECURITY_DELAY','AIRLINE_DELAY','LATE_AIRCRAFT_DELAY','WEATHER_DELAY']]
y=df['result']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=1)
```

```python
ss = StandardScaler().fit_transform(x_train, x_test)
```

```python
clf = DecisionTreeClassifier()
clf = clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict_proba(x_test)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score = roc_auc_score(y_test, y_pred[:,1])
roc_auc_score
```
