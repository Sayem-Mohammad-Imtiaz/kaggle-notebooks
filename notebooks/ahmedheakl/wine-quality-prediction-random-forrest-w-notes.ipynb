````python
X = df.drop(columns='quality', axis=1)
y = df['quality'].apply(lambda x: 1 if x > 6 else 0).values
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=True, stratify=y, random_state=0)
````

````python
print(f'X_train: {X_train.shape}')
print(f'X_test: {X_test.shape}')
````

````python
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
````

````python
print('Training Accuracy: ', classifier.score(X_train, y_train))
print('Testing Accuracy: ', classifier.score(X_test, y_test))
````

The model is clearly overfitting since the training accuracy is ```1.0```. But it is ok, since the test acc is also high. 

````python
def predict(X, model):
    X = np.array(X).reshape(1, -1)
    pred = model.predict(X)
    if pred == 1:
        return 'Good Wine Quality'
    else:
        return 'Bad Wine Quality'
````

````python
predict(X_test.values[3], classifier)
````
