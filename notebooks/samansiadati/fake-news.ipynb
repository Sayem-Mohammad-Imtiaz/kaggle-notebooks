```python
import pandas as pd
```

```python
fake = pd.read_csv('../input/fake-and-real-news-dataset/Fake.csv', delimiter = ',')
fake['sentiment']= 0
fake
```

```python
true = pd.read_csv('../input/fake-and-real-news-dataset/True.csv', delimiter = ',')
true['sentiment']= 1
true
```

```python
df =pd.DataFrame()
df = true.append(fake)
column = ['date','subject']
df = df.drop(columns=column)
```

```python
import numpy as np
input_array=np.array(df['title'])
```

```python
import re
import nltk
# ltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
corpus = []
for i in range(0, 40000):
    review = re.sub('[^a-zA-Z]', ' ', input_array[i])
    review = review.lower()
    review = review.split()
    ps = PorterStemmer()
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000)
X = cv.fit_transform(corpus).toarray()
y = df.iloc[0:40000, 2].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
