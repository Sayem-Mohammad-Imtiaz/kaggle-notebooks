# Prediction

### Data Preparation

```python
from sklearn import preprocessing
```

```python
label_encoder = preprocessing.LabelEncoder()
```

Sex

```python
Sex = label_encoder.fit_transform(df['sex'])
Sex = pd.DataFrame({'sex': Sex})
```

```python
df.sex = Sex
```

```python
df.sex 
```

Smoker

```python
Smoker = label_encoder.fit_transform(df['smoker'])
Smoker = pd.DataFrame({'smoker': Smoker})
```

```python
df.smoker = Smoker
```

```python
df.smoker 
```

Region

```python
Region = label_encoder.fit_transform(df['region'])
Region = pd.DataFrame({'region': Region})
```

```python
df.region = Region
```

```python
df.region 
```

```python
y = df.charges 
```

```python
X = df.drop(columns =['charges'])
```

```python
X.head(10)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
X = pd.DataFrame(X)
```

```python
X
```

### Data Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=1)
```

### Linear regression

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor_linear = LinearRegression()
```

```python
regressor_linear.fit(X_train, y_train)
```

```python
y_predict = regressor_linear.predict(X_test)
```

```python
y_predict
```

```python
Prediction = pd.DataFrame({'charges': y_predict})
```

```python
Prediction
```

```python
from sklearn.metrics import mean_squared_error
```

```python
import numpy as np
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_predict))
print("RMSE: %f" % (rmse))
```

### XGB Regression

```python
from xgboost import XGBRegressor
```

```python
regressor_XGB = XGBRegressor(n_estimators=1000, learning_rate=0.05)
```

```python
regressor_XGB.fit(X_train, y_train, early_stopping_rounds=5, eval_set=[(X_test, y_test)], verbose=False)
```

```python
y_predict = regressor_XGB.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_predict))
print("RMSE: %f" % (rmse))
```
