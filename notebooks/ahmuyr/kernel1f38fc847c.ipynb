```python
grocery_list = pd.read_csv('/kaggle/input/supermarket/GroceryStoreDataSet.csv',header=None, names=['items'])
```

```python
df = grocery_list.copy()
```

```python
df.head()
```

```python
df.values
```

```python
data = list(df['items'].apply(lambda x:x.split(',')))
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
df1 = apriori(df, min_support=0.2, use_colnames=True)
```

```python
df1.sort_values(by="support",ascending=False)
```

```python
df1['length'] = df1['itemsets'].apply(lambda x:len(x))
df1
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(df1, metric = "confidence", min_threshold = 0.2)
```

```python
rules = association_rules(df1, metric = "lift", min_threshold = 1)
rules
```

2. Ekmek ve sütün ayni alisveriste alinma orani 80% ve lift 1.2 yani bu bize bir cross-selling olasiligi sagliyor. Musterinin ekmek ve sutu alacagindan emin oldugumuz icin aralarini oldukca acip musteriyi baska reyonlardan gecirip daha cok satis yapabiliriz.
3. Hazir corba alan kisilerin cay alma olasiligi %80. Cay alan kisilerin ise hazir corba alma olasiligi %57 altinda. Bir musterinin cay alma olasiligi %35, hazir corba alma olasiligi ise %25. Bu ihtimalleri göz önune aldigimizda karsimiza iki olasilik cikiyor. Musteri ilk önce cay alirsa hazir corba alma olasiligi: 0.35 * 0.57 = 0.1995, Musteri ilk once hazir corba alirsa yaninda cay alma ihtimali ise: 0.20 * 0.8 = 0.16. Bu yuzden cayin reyonunun hazir corba reyonundan gelmesi bizim icin daha karli olacaktir.
