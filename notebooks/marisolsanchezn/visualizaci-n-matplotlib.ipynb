# Gráficas

```python
#importamos librería para gráficos
import matplotlib.pyplot as plt
```

 A. Histogramas

```python
#histograma frecuencia de la edad
plt.hist(datos.age)
```

```python
#histograma frecuencia de la edad
plt.hist(datos.age, density = True, color = 'green')
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.title("Histograma de frecuencia relativa de la edad")
```

```python
plt.hist(datos.bmi, density = True, color = "red", edgecolor = "black")
plt.title('Histograma de frecuencia relativa del bmi')
plt.xlabel('bmi') 
plt.ylabel('frecuncia') 
plt.legend('bmi') 
```

```python
plt.hist(datos.charges,density = True,edgecolor = 'black',color="purple")
plt.xlabel("Cargos")
plt.ylabel("Frecuencia")
plt.title("Histograma de frecuencia relativa de los cargos")
plt.axvline(datos.charges.mean(), color='blue', label = "media")
plt.legend()
```

```python
plt.hist(datos.children,density = True,edgecolor = 'black',color="orange")
plt.xlabel("Hijos")
plt.ylabel("Frecuencia")
plt.title("Histograma de frecuencia relativa de número de hijos")
plt.axvline(datos.children.mean(), color='blue', label = "media")
plt.axvline(datos.children.median(), color='green', label = "media")
plt.legend()
```

```python
plt.hist(datos.region,color="pink")
plt.xlabel("Region")
plt.ylabel("Frecuencia")
plt.title("Histograma de frecuencia relativa de regiones")
```

B. Gráfica de Pie

```python
grupos = datos.groupby('sex').count()
lista = grupos.age
lista

plt.pie(lista,autopct='%1.2f%%')
plt.title("Sexo")
plt.legend(lista.index)
```

```python
grupos = datos.groupby('region').count()
lista = grupos.age
lista

plt.pie(lista,autopct='%1.2f%%', explode =(0, 0, 0.1, 0), shadow = True)
plt.title("Regiones")
plt.legend(lista.index)
```

```python
grupos = datos.groupby('smoker').count()
lista = grupos.age
lista

plt.pie(lista,autopct='%1.2f%%')
plt.title("Fumador")
plt.legend(lista.index)
```

```python
grupos = datos.groupby('children').count()
lista = grupos.age
lista

plt.pie(lista,autopct='%1.2f%%')
plt.title("Número de hijos")
plt.legend(lista.index) 
```

C. Gráfico lineal

```python
meses = ["Ene","Feb", "Mar", "Abr", "May", "Jun", "Jul",
         "Ago", "Sep", "Oct", "Nov", "Dic" ]
ventas = [750, 930, 1272, 1072, 1135, 1486, 1438, 1106, 1339, 1449, 1551, 1761]

plt.plot( meses , ventas)
plt.title("Ventas mensuales 2019")
plt.xlabel("Mes")
plt.ylabel("Ventas")
```

D. Gráfica de Barras

```python
grupos = datos.groupby('region').mean()
regiones = grupos.index
cargos_promedio = grupos.charges

plt.bar(regiones,cargos_promedio)
plt.title("Cargo promedio por Región")
plt.xlabel("Región")
plt.ylabel('Cargo Promedio')
```

E. Gráfica de Disperción

```python
plt.scatter(datos.age,datos.charges, color = 'red')
plt.title("Gráfica de Edad vs Cargos")
plt.xlabel("Edad")
plt.ylabel('Cargos')
```
