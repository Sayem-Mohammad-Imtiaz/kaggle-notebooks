```python
# Choose any numerical column
freedom = df['Freedom']
```

```python
# Graph Distribution Plot of Column
sns.distplot(freedom,kde=False,bins=10,hist_kws={'lw':1,'edgecolor':'gray'}).set_title('Freedom Distribution Plot')
plt.grid()
```
