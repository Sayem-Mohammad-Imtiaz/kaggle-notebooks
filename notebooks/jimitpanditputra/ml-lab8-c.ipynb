# SVM

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../input/iris-flower-dataset/IRIS.csv')
df
```

```python
y = df.iloc[:,-1]
y1=y.unique()
y1
```

```python
species_num= {'species':{'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2}}
species_num
```

```python
df.replace(species_num, inplace=True)
```

```python
df
```

```python
df.shape
```

```python
df['names'] =df.species.apply(lambda x: y1[x])
df.head()
```

```python
x = df.iloc[:,:2]
y = df.iloc[:,-2]
```

```python
plt.scatter(x['sepal_length'],x['sepal_width'],c=y,cmap=plt.cm.Paired)
```

```python
from sklearn.svm import SVC
```

```python
classifier = SVC(kernel='linear' )
classifier.fit(x,y)
```

```python
classifier.score(x,y)
```

```python
x_min, x_max = df.iloc[:, 0].min() , df.iloc[:, 0].max() 
y_min, y_max = df.iloc[:, 1].min() , df.iloc[:, 1].max() 
h = (x_max / x_min)/100
x1, y1 = np.meshgrid(np.arange(x_min, x_max, h),
np.arange(y_min, y_max, h))
```

```python
plt.subplot(1, 1, 1)
Z = classifier.predict(np.c_[x1.ravel(), y1.ravel()])
Z = Z.reshape(x1.shape)
plt.contourf(x1, y1, Z, cmap=plt.cm.Paired, alpha=0.8)
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=y, cmap=plt.cm.Paired)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')

plt.show()
```

```python
classifier1 = SVC(kernel='rbf', C=1, ).fit(x,y)
classifier2 = SVC(kernel='poly', C=1, ).fit(x,y)
classifier3 = SVC(kernel='sigmoid', C=1, ).fit(x,y)
```

```python
titlr=['rbf','polynomial','sigmoid']
for i, classifier in enumerate((classifier1,classifier2,classifier3)):
    plt.subplot(2, 2, i + 1)
    plt.subplots_adjust(wspace=0.1, hspace=0.1) 
    Z = classifier.predict(np.c_[x1.ravel(), y1.ravel()])
    Z = Z.reshape(x1.shape)
    plt.contourf(x1, y1, Z, cmap=plt.cm.Paired, alpha=0.8)
    plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=y, cmap=plt.cm.Paired)
    plt.xlabel('Sepal length')
    plt.ylabel('Sepal width')

    plt.title(titlr[i])
    plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.2, )
classifier = SVC(kernel='linear', C=1, ).fit(x_train,y_train)
classifier1 = SVC(kernel='rbf', C=1, ).fit(x_train,y_train)
classifier2 = SVC(kernel='poly', C=1, ).fit(x_train,y_train)
classifier3 = SVC(kernel='sigmoid', C=1, ).fit(x_train,y_train)
```

```python
#Linear
classifier.score(x_test,y_test)
```

```python
#Radial basis function
classifier1.score(x_test,y_test)
```

```python
#polynomial
classifier2.score(x_test,y_test)
```

```python
#sigmoid
classifier3.score(x_test,y_test)
```
