```python
import pandas  as pd #Data manipulation
import numpy as np #Data manipulation
import matplotlib.pyplot as plt # Visualization
import seaborn as sns #Visualization
```

```python
df=pd.read_csv("../input/insurance/insurance.csv")
df
```

#Exploratory Data Analysis

```python
df.describe()
```

```python
df.info()
```

Hence there are no null values or missing places to be managed

```python
from sklearn import linear_model
lin=linear_model.LinearRegression()
```

```python
plt.scatter(df.age,df.charges)
plt.xlabel("Age")
plt.ylabel("Charges")
lin.fit(df[['age']],df.charges)
plt.plot(df.age,lin.predict(df[['age']]),color='red')
lin.predict([[40]])
```

```python
plt.scatter(df.bmi,df.charges)
plt.xlabel("BMI")
plt.ylabel("Charges")
lin.fit(df[['bmi']],df.charges)
plt.plot(df.bmi,lin.predict(df[['bmi']]),color='red')
lin.predict([[35]])
```

```python
plt.scatter(df.children,df.charges)
plt.xlabel("Children")
plt.ylabel("Charges")
lin.fit(df[['children']],df.charges)
plt.plot(df.children,lin.predict(df[['children']]),color='red')
lin.predict([[2]])
```

**Multiple Variable**

```python
lin.fit(df[['age','bmi','children']],df.charges)
```

```python
lin.predict([[40,35,2]])    #predicting value based on linear regression
```
