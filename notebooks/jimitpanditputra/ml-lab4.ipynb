```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

```python
df=pd.read_csv("../input/pima-indians-diabetes-database/diabetes.csv")
df
```

**Splitting independent and dependent variables**

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
```

```python
y.unique()
```

**This represents the simple unique outputs available**

**Now Splitting training and testing data   </br>
Where the test data size is 10% of the total data**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
print("shape of x train=",x_train.shape)
print("shape of x test=",x_test.shape)
print("shape of y train=",y_train.shape)
print("shape of y test=",y_test.shape)
```

# Classificatoin Model

```python
from sklearn.tree import DecisionTreeClassifier
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(x_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(x_test)
```

```python
#Checking the accuracy of the model
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
#Total Depth of the tree which ususally leads to overfitting
print(clf.get_depth())
```

# Optimizing Decision Tree Performance

**Using Entropy Classifier </br>
Using max_depth parameter to assign value to prevent overfitting or underfitting**

```python
clf2 = DecisionTreeClassifier(criterion="entropy", max_depth=3)
clf2 = clf.fit(x_train,y_train)
y_pred = clf2.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor(criterion='mse')
reg=reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor(criterion='mae')
reg=reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

**Trying to optimize the regressor**

```python
from sklearn.tree import DecisionTreeRegressor
reg=DecisionTreeRegressor(criterion='mae',max_depth=3)
reg=reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
