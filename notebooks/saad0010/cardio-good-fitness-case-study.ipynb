# Dataset Description

The market research team at AdRight is assigned the task to identify the profile of the typical customer for each treadmill product offered by CardioGood Fitness.

Product: There are three types of treadmills mentioned in the dataset.
Age: Age of the customer in years.
Gender: Gender of the customer.
Education: Education og the customer in years.
MaritalStatus: whether the customer is single or has a partner.
Usage: Average number of times the treadmill is expected to be used per week.
Income: Annual household income in $.
Fitness: Self rated fitness of customer from 0 to 5. 0 unfit and 5 fit.
Miles: Average number of miles cutomer expects to walk/run each week.

# Analysing relation among the features that affect choice of product

How gender and marital status affect the choice of product

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Product',hue='Gender',data=df);
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Product',hue='MaritalStatus',data=df);
```

* Gender does not affect the choice of product in case of 'TM195' and 'TM498' whereas 'TM798' has significantly large male customers as compared to female customers. Females tend to prefer 'TM195' and 'TM498'.

* Marital Status does not affect the choice of the treadmill.
