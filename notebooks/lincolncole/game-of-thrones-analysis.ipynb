# Let's Take A Look At This Game Of Thrones Data

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sn # data visualization
%matplotlib inline
sn.set (color_codes = True, style="white")
import matplotlib.pyplot as ml #data visualization
import warnings
warnings.filterwarnings("ignore")

print ("Environment works!!!")
```

```python
# Let's import our data 
deaths = pd.read_csv("../input/character-deaths.csv", sep = ",", header = 0)
```

```python
# Let's take a look at the shape and size of our data

deaths.shape
```

```python
# Let's take a look at this data

deaths.head()
```

```python
# Let's get rid of any rows where "NaN" is a value. This should leave us with the characters that have died.

deaths.dropna(axis=0, how='any')
```

```python
# Let's plot a histogram

deaths.hist()
ml.show()
```

```python
# Let's plot a box plot

sn.boxplot(data=deaths)
```

```python
# Let's look at a violin plot comparing deaths with gender

sn.violinplot(x="Gender", y="Death Year", data=deaths)
```

# Males started to be killed earlier on in the series.
# The rate at which male characters get killed is less consistent than that of female characters.

```python
deaths.Gender.value_counts()
```

```python
# Let's look at a violin plot comparing deaths with nobility

sn.violinplot(x="Nobility", y="Death Year", data=deaths)
```

# There appears to be very little correlation between deaths and nobility.

```python
# Lets have a look at the heatmap

co = deaths.corr()
sn.heatmap(co, annot=True, linewidths=1.0)
```

# There appears to be a low correlation between being noble and the chance of dying and an even lower correlation between gender and the chance of dying.
