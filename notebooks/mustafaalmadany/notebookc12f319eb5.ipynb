```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
Input_Path= '/kaggle/input/netflix-shows/netflix_titles.csv'
```

```python
raw_df=pd.read_csv(Input_Path)
```

```python
raw_df
```

```python
raw_df.dtypes
```

```python
df=raw_df.copy()
df['date_added']=pd.to_datetime(df['date_added'])
```

```python
df.dtypes
```

```python
show_number_per_day=df.groupby(['date_added'])[['show_id']].count()
show_number_per_day
```

**Extract the day name and sum-up the shows added**

```python
#Extract the name of day and sum-up the show in every day
show_number_per_day['day_of_week']=show_number_per_day.index.day_name()#here we creat a new column "day_of_week" 
show_number_per_day
```

```python
#Now we need to groupby the series I create to sum the number of show added per day
show_added_per_day=show_number_per_day.groupby(['day_of_week'])[['show_id']].sum()
show_added_per_day.rename(columns={'show_id':'Num_of_show_added'},inplace=True)
show_added_per_day
```

```python
# Select the data from 2016 and forward
date_filter=show_number_per_day.index >= '2016-01-01'
show_filtered=show_number_per_day[date_filter]
show_filtered
```

```python
#Sum up the weekly data from 2016 and forward
show_per_week=show_filtered[['show_id']].resample('1W').sum()
show_per_week
```

```python
# Find the sum of show added by year and make some charts to show how netflix develop by years with movie and series
show_df=df[['show_id','type','date_added']]
show_df.set_index('date_added',inplace=True)
movie_count=show_df[show_df['type']=='Movie']
movie_num_per_year=movie_count.resample('Y').count()
movie_num_per_year
```

```python
movie_num_per_year.drop(columns=['type'],inplace=True)
movie_num_per_year.rename(columns={'show_id':'movie(s)_added_byYear'},inplace=True)
movie_num_per_year
```

```python
Tv_show_count=show_df[show_df['type']=='TV Show']
Tv_show_num_per_year=Tv_show_count.resample('Y').count()
Tv_show_num_per_year.drop(columns=['type'],inplace=True)
Tv_show_num_per_year.rename(columns={'show_id':'shows_added_per_yaer'},inplace=True)
Tv_show_num_per_year
```
