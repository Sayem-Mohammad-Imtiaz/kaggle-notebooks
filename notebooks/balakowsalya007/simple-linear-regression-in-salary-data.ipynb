**Simple Linear Regression**

**Import packages/libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

**Loading dataset**

```python
salaryData = pd.read_csv('../input/salary-data-simple-linear-regression/Salary_Data.csv')
salaryData.head()
```

**Data Summary**

```python
salaryData.describe()
```

```python
salaryData.shape
```

```python
salaryData.isnull().sum()
```

```python
X = salaryData.iloc[:,:-1]
y = salaryData.iloc[:,-1]
```

```python
print(X)
print(y)
```

**Spliting the data into test and train set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state = 111, test_size = 0.3)
```

**Library itself will apply scaling function**

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

****Visualising the training set results****

```python
plt.scatter(X, y, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary') 
plt.show()
```
