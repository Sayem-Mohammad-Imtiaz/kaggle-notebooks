### Start Time

```python
var.start_time = pd.to_datetime(var.Start_Time)
```

```python
var.start_time = pd.to_datetime(var.start_time)
```

```python
sns.distplot(var.start_time.dt.hour, bins = 24, kde = False , norm_hist = True)
```

### High percent of accident happen between 6 am to 10 am.
### Next from 3 pm to 6pm

```python
sns.distplot(var.start_time.dt.dayofweek, bins = 7, kde = False , norm_hist = True)
```

### Accidents are high on weekdays because low number of people use vehicle on weekends.

```python
sundays = var.start_time[var.start_time.dt.dayofweek ==6]
```

```python
sns.distplot(sundays.dt.hour, bins = 24, kde = False , norm_hist = True)
```

### On weekends accident are in afternoon.

```python
monday = var.start_time[var.start_time.dt.dayofweek ==1]
sns.distplot(monday.dt.hour, bins = 24, kde = False , norm_hist = True)
```

### Graph show 2 Gausian curve . 
##### 1) People going to work
##### 2) People returning from work

```python
sns.distplot(var.start_time.dt.month, bins = 12, kde = False , norm_hist = True)
```
