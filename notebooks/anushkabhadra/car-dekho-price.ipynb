```python
def process_name(car_name):
    
    return car_name.split()[0]
```

```python
df['brand'] = df['name'].apply(process_name)
```

```python
df.drop('name',axis=1,inplace=True)
```

```python
df['year_old'] = 2020 - df['year']
```

```python
df.drop('year',axis=1,inplace=True)
```

```python
df['fuel'].value_counts()
```

```python
df = df[~df['fuel'].isin(['CNG','LPG','Electric'])]
```

```python
df = df[~(df['seller_type'] == 'Trustmark Dealer')]
```

```python
df = df[~df['owner'].isin(['Fourth & Above Owner','Test Drive Car'])]
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

fuel_encoder = LabelEncoder()
seller_encoder = LabelEncoder()
transmission_encoder = LabelEncoder()
owner_encoder = LabelEncoder()
brand_encoder = LabelEncoder()
```

```python
df['fuel'] = fuel_encoder.fit_transform(df['fuel'])
df['seller_type'] = seller_encoder.fit_transform(df['seller_type'])
df['transmission'] = transmission_encoder.fit_transform(df['transmission'])
df['owner'] = owner_encoder.fit_transform(df['owner'])
df['brand'] = brand_encoder.fit_transform(df['brand'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first')
```

```python
X_temp = ohe.fit_transform(df[['owner','brand']]).toarray()
```

```python
X_temp.shape
```

```python
X = df.drop(['owner','brand','selling_price'],axis=1).values
```

```python
X.shape
```

```python
X = np.hstack((X,X_temp))
```

```python
X
```

```python
y = df.iloc[:,0].values
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(n_estimators=500)
```

```python
rf_reg.fit(X_train,y_train)
```

```python
y_pred1 = rf_reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error

print(r2_score(y_test,y_pred1))
print(mean_absolute_error(y_test, y_pred1))
```
