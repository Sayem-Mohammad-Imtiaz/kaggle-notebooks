**What ingredients or flavors are most commonly advertised on ramen package labels?**

Extract most common flavor using keyword extraction

**Lowercasing** all the text data, it can help in cases where dataset is 
not very large and significantly helps with consistency of expected output.

```python
df['Variety'] = df['Variety'].str.lower()
```

```python
#remove special characters
special_characters ='\W+'
df['Variety'] = df['Variety'].str.replace(special_characters, ' ', regex=True)
```

```python
#word token
from nltk.tokenize import sent_tokenize,word_tokenize

df['Variety'] = df['Variety'].apply(word_tokenize)
```

```python
#Remove stopword
import nltk
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

df['Variety'] = df['Variety'].apply(lambda text: [w for w in text if w not in stop_words])
```

```python
#Lemmatization
from nltk.stem.wordnet import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

df['Variety'] = df['Variety'].apply(lambda text: list(set([lemmatizer.lemmatize(w) for w in text])))
```

```python
#all keywords
all_keywords =[]
for each_lst in df['Variety']:
    all_keywords.extend(each_lst)

#
import nltk 
lst = nltk.pos_tag_sents([all_keywords])[0]
keywords = []
for wd,tag in lst:
    if tag not in ['RB','RP','VBZ','CD','CC']:
        keywords.append(wd)
from collections import Counter
keyword_dict = Counter(keywords)
```

```python
most_common_ingredient = dict(sorted(keyword_dict.items(), key=lambda item: item[1],reverse=True))
most_common_ingredient
```
