```python
data = pd.read_csv("../input/stock-market-data/stock_market_data/nasdaq/csv/AAPL.csv")
data.tail()
```

```python
#pre-pro
from sklearn.impute import SimpleImputer

imputer = SimpleImputer()
data = data[['Low', 'Open', 'Volume', 'High', 'Close', 'Adjusted Close']]
imputed_data = pd.DataFrame(imputer.fit_transform(data))
imputed_data.columns = data.columns
data = imputed_data
```

```python
from sklearn.model_selection import train_test_split
X = data[['Low', 'Open', 'Volume', 'Close', 'Adjusted Close']]
y = data.High

X_train, X_valid, y_train, y_valid = train_test_split(X, y)
```

```python
from sklearn import metrics 
from sklearn.linear_model import SGDRegressor,Lasso
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error,auc,confusion_matrix

model = Lasso(alpha=1.4,max_iter=2500)
model.fit(X_train, y_train)

pred = model.predict(X_valid)
print('Mean Absolute Error :',mean_absolute_percentage_error(y_valid, model.predict(X_valid)))
#fpr, tpr = metrics.roc_curve(y, pred)
#print(len(y))
#print(len(pred))
#print(metrics.confusion_matrix(y_valid,pred))
```
