```python
f,ax = plt.subplots(figsize=(14,14))
sns.heatmap(data_2017.corr(), annot=True, linewidths=.5, fmt='.2f',ax=ax, cmap="YlGn")
```

Above heatmap shows the correlation. What can we say by looking the map above.

Well lets take 'Hapiness Score' as key column and look it's correlation with others.

* Correlation between *Hapiness Score* and *GDP per Capita* is **0.81**
* Correlation between *Hapiness Score* and *Health..Life.Expectancy* is **0.78**
* Correlation between *Hapiness Score* and *Family* is **0.75**
* Correlation between *Hapiness Score* and *Freedom* is **0.57**
* Correlation between *Hapiness Score* and *Dystopia.Residual* is **0.48**
* Correlation between *Hapiness Score* and *Trust..Government.Corruption* is **0.43**
* Correlation between *Hapiness Score* and *Generosity* is **0.16**

**By looking these information, it's clear that GDP per Capita has highest correlation with Hapiness Score while Generosity has lowest.**

Let's look freedom's general view on Happiness Rank.

```python
data_2015.plot(kind='scatter', x='Happiness Rank', y='Freedom',alpha = 0.9,color = 'green')
plt.xlabel('Happiness Rank')              # label = name of label
plt.ylabel('Freedom')
plt.title('Happiness Rank over Freedom Scatter Plot')   
```

From the top left corner of scatter diagram, it can be gathered that top ranked countries have high freedom points in general, as expected. 

In the top-right corner, there is a country with high freedom point but ranked in last 20 in Hapiness. Let's find out which country that is, just wondered :)
