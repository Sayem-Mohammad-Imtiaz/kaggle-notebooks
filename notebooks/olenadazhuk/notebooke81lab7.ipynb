Найдите средний балл студентов школы GP моложе 19 лет.

```python
sd= df[df["school"] == "GP"]
print(sd)
sm= sd[sd['age']< 19]
sm['G3'].mean()
```

Постройте инфографику зависимости среднего балла (G3) от количества времени,
уделяемого учёбе (studytime).

```python
sns.relplot(x='G3', y='studytime', kind='line',data=df)
```

Отобразите на графике соотношение между количеством пропусков (absences)
и оценкой (G3). Вычислите коэффициент корреляции (pd.Series.corr()).

```python
sns.relplot(x='absences', y='G3',kind='line',color='green',data=df)
```
