### Import the basic liberaries and import the dataset

```python
# Import Basic liberaries
import numpy as np
import pandas as pd 
pd.set_option('display.max_columns' , 500)

import seaborn as sns
import matplotlib.pyplot as plt

#Read the data
df = pd.read_csv('/kaggle/input/cost-of-living/cost-of-living.csv')
df.head()
```

### The data shows the cost of every things across the cities  , <br><br> MacMeal in Egypt cost 4.5 usd however in Italy cost 8 usd <br><br> I will sum the cost of every thing and assign it to the countries <br><br> then we will sort the countries from the most expensive to the cheapest

<h2 style = 'background-color:lightgreen;border-style: solid'><center><strong><br> Re-Shaping the data

```python
#Here goes the data manupilation pipeline
df = df.T #Transpose the dataset , put the cities in rows
new_header = df.iloc[0] #grab the first row for the header
df = df[1:] #take the data less the header row
df.columns = new_header #set the header row as the df header
df['Total_cost'] = df.sum(axis=1) #generate column for total cost 
df = df.reset_index()
#Extract the country column from the location
df.rename(columns  = {'index' : 'City'} , inplace = True)
df[['City' , 'Country' , 'other']] =df['City'].str.split(',', expand=True)
#Deal with the states of United states
state = df[df.other == ' United States']['Country'].tolist()
df.Country = df.Country.apply(lambda x: 'United State' if x in state else x)
df.drop('other' , axis = 1 , inplace = True)
#Combine the data by the country and sort the average total cost
df_countries = df.groupby('Country')['Total_cost'].mean().reset_index().sort_values('Total_cost' ,axis=0, ascending=False, ignore_index=True )
```

<h2 style = 'background-color:Violet;border-style: solid'><center><strong> <br>Where your Counry locate in the chart?

```python
fig , ax  = plt.subplots(figsize = (10 , 30))
sns.barplot(data = df_countries , y = 'Country' , x = 'Total_cost')
plt.title('Sorting the most expensive Countries' , fontsize  = 14);
plt.grid(axis = "x")
for s in ['top', 'right', 'bottom']:
    ax.spines[s].set_visible(False)
    
for patch in ax.patches:
    bl = patch.get_xy()
    x = 0.9 * patch.get_width() + bl[0]
    y = 0.5 * patch.get_height() + bl[1] 
    ax.text(x,y,"%d" %(patch.get_width()),va='center'
            ,ha='center' , color = '#4a4a4a',
           fontsize = 12, fontfamily='serif')
```
