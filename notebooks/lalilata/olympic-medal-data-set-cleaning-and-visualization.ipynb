### Graph 1: Countries with most medals

```python
medal_sort = df_summer.Country.value_counts() # value_counts() return a Series containing counts of unique values
medal_sort
```

```python
df_medal = medal_sort.to_frame()
df_medal.head(n = 10)
```

```python
df_medal = df_medal.rename(columns = {"country": "Code", "Country": "medal_count"})
```

```python
df_medal['Code'] = df_medal.index
```

```python
df_medal.head()
```

```python
country = ["USA", "URS", "GBR", "FRA", "GER", "ITA", "AUS", "HUN", "SWE", "NED"]
medal = [4585, 2049, 1720, 1396, 1305, 1296, 1189, 1079, 1044, 851]
```

```python
from matplotlib import pyplot as plt

figure(figsize=(12, 6), dpi=80)

plt.bar(country, medal) 
plt.xlabel("Medal Count")
plt.ylabel("Country Name")
plt.title("Top 10 Country with most Olympic medals")
plt.show()
```

### Graph 2: Correlation of number of Medals to GDP per capita

##### Let's find out does countries earn more medals by making more money!

```python
df_dictionary.head()
```

```python
df_forgdp = df_dictionary[['Country', 'Code', 'GDP per Capita']]
df_forgdp.head()
```

```python
df_medal.head()
```

```python
df_medal_gdp = pd.merge(df_forgdp, df_medal, on = "Code", how = 'left') #merge dataframes in a SQL-like fashion
```

```python
df_medal_gdp.head()
```

```python
df_medal_gdp = df_medal_gdp.rename(columns = {"GDP per Capita": "GDP_per_Capita"})
```

```python
x = df_medal_gdp.GDP_per_Capita
y = df_medal_gdp.medal_count
```

```python
plt.scatter(x, y) 
plt.xlabel("GDP per Capita")
plt.ylabel("Medal count")
plt.title("More money means more Medals?")
plt.show()
```
