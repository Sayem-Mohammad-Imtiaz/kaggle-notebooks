```python
# Import essential libraries
from matplotlib import pyplot as plt
import seaborn as sns
from matplotlib import cm
import pylab as pl
from sklearn import linear_model, metrics, model_selection
from sklearn.decomposition import PCA
from pandas.plotting import scatter_matrix, andrews_curves, parallel_coordinates, lag_plot, autocorrelation_plot, bootstrap_plot, radviz, table
```

```python
# Reading the dataset
data_2015 = pd.read_csv('../input/world-happiness/2015.csv')
data_2016 = pd.read_csv('../input/world-happiness/2016.csv')
data_2017 = pd.read_csv('../input/world-happiness/2017.csv')
data_2018 = pd.read_csv('../input/world-happiness/2018.csv')
data_2019 = pd.read_csv('../input/world-happiness/2019.csv')
```

```python
score_2019 = np.array(data_2019['Score'])
GDP_2019 = np.array(data_2019['GDP per capita'])
```

```python
#Finding correlation between GDP and Happiness Score
corr = data_2019.corr(method='pearson')
corr.head()
```

```python
sns.heatmap(corr)
```

```python
#linear Regression
plt.scatter(GDP_2019, score_2019)
plt.xlabel('GDP per capita of 2019')
plt.ylabel('happiness score of 2019')
plt.show()
```

```python
# Fit the model, determine regression formula y = 2.218148x + 3.399345178292416 as well as MSE 0.4551967397324353
score_model = linear_model.LinearRegression(normalize=True)
x = np.expand_dims(GDP_2019, 1)
y = score_2019
score_model.fit(x, y)

#MSE calculation
train_pred = score_model.predict(x)
mse_train = metrics.mean_squared_error(y, train_pred)
print(mse_train)

a = score_model.coef_
b = score_model.intercept_
print(score_model.coef_)
print(score_model.intercept_)
```

```python
#visualize regression
plt.scatter(GDP_2019, score_2019)
x = np.linspace(0, 2)
y = a*x + b
plt.plot(x, y, 'r')
plt.xlabel('GDP per capita of 2019')
plt.ylabel('happiness score of 2019')
plt.show()
```

```python
#residual plot to find outliers
sns.residplot(x=GDP_2019, y=score_2019, scatter_kws=dict(s=50))
tolerant_distance = 1.4

# annotations indicating an outlier
#plt.annotate('Botswana', xy=(1.041,3.488), xytext=(1.05,22), fontsize=14,
#            arrowprops=dict(arrowstyle='->', ec='grey', lw=2), bbox = dict(boxstyle="round", fc="0.8"))

# ticks
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# labels and title
plt.xlabel('GDP per capita of 2019', fontsize=14)
plt.ylabel('residuals', fontsize=14)
plt.axhline(y=tolerant_distance, color='r', linestyle='-')
plt.axhline(y=-tolerant_distance, color='r', linestyle='-')
plt.title('Residual plot', fontsize=20);




#Two data point with tolerant distance of 1.4: 
#Botswana(one at the bottom) with happniness score 3.488 and GDP 1.041
#Costa Rica(one at the top) with happniness score 7.167 and GDP 1.034
```
