```python
# Quitar Nans, sustituir por media de cada columna.
lista_cuantitativa = list()
lista_cualitativa = list()
for i in df.columns[4:]:
    df[i] = df[i].replace(np.nan,df[i].mean())
    lista_cuantitativa.append(i)
for i in df.columns[:4]:
    lista_cualitativa.append(i)        
```

- Define el rango de cada uno de los atributos (cuantitativos)

```python
rango_pf_rol_procedural = np.max(df["pf_rol_procedural"])-np.min(df["pf_rol_procedural"])
rango_pf_rol_procedural
```

```python
lista_rangos = list()
for i in df.columns[4:]:
    lista_rangos.append(np.max(df[i])-np.min(df[i]))
lista_rangos
```

- Obtén la moda o promedio dependiendo el caso para cada atributo

```python
import scipy
from scipy import stats 

#MODA CUANTITATIVOS
lista_moda = list()
for i in df.columns[4:]:
    lista_moda.append(stats.mode(np.array(df[i])))
lista_moda
```

```python
#PROMEDIO CUANTITATIVOS
lista_promedio = list()
for i in df.columns[4:]:
    lista_promedio.append(np.mean(np.array(df[i])))
lista_promedio
```

```python
#MODA CUALITATIVOS
lista_moda_cualitativos = list()
for i in df.columns[:4]:
    lista_moda_cualitativos.append(stats.mode(np.array(df[i])))
lista_moda_cualitativos
```

- Obtén el rango, varianza, desviación estandar para los atributos cuantitativos.

```python
#RANGO HECHO ARRIBA
#VARIANZA

lista_varianza = list()
for i in df.columns[4:]:
    lista_varianza.append(np.var(np.array(df[i])))
lista_varianza
```

```python
# DESVIACION ESTANDAR
lista_desv_estandar = list()
for i in df.columns[4:]:
    lista_desv_estandar.append(np.std(np.array(df[i])))
lista_desv_estandar
```
