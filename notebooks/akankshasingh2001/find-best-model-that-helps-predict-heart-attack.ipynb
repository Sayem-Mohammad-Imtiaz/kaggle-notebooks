**FITTING DATA TO THE MODEL**

```python
x = train.iloc[:,:-1].values
y = train.iloc[:,-1].values
print(x)
print(y)
```

```python
x_train,x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 23)
```

```python
sc = StandardScaler()
sc.fit_transform(x_train)
```

LOGISTIC REGRESSION

```python
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
y_pred_lr = log_reg.predict(x_test)
print("Accuracy of Logistic Regression is", accuracy_score(y_test, y_pred_lr))
```

SUPPORT VECTOR MACHINES

```python
svc = SVC()
svc.fit(x_train, y_train)
y_pred_svc = svc.predict(x_test)
print("Accuracy of Support Vector Machine is", accuracy_score(y_test, y_pred_svc))
```

Hyperparameter Tuning

```python
svm = SVC()
parameter = [{'kernel': ['rbf'], 'gamma': [1e-3, 1e-4],
              'C': [1, 10, 100, 1000]},
             {'kernel': ['linear'],
              'C': [1, 10, 100, 1000]}]

searcher = GridSearchCV(svm, parameter)

searcher.fit(x_train, y_train)

print("The best set of parameters for SVM are:", searcher.best_params_)
print("The best score of SVM is:", searcher.best_score_)

y_pred_svm = searcher.predict(x_test)

print("Accuracy score of SVM after hyperparameter tuning is:", accuracy_score(y_test, y_pred_svm))
```

DECISION TREE CLASSIFIER

```python
dt = DecisionTreeClassifier(random_state = 26)
dt.fit(x_train, y_train)

y_pred_dt = dt.predict(x_test)

print("Accuracy score of Decision Tree is:", accuracy_score(y_test, y_pred_dt))
```

RANDOM FOREST CLASSIFIER

```python
rf = RandomForestClassifier(n_estimators = 40, random_state =  61)
rf.fit(x_train, y_train)

y_pred_rf = rf.predict(x_test)

print("Accuracy score of Random Forest is:", accuracy_score(y_test, y_pred_rf))
```

GRADIENT BOOST CLASSIFIER

```python
gb = GradientBoostingClassifier(random_state = 31)
gb.fit(x_train, y_train)

y_pred_gb = gb.predict(x_test)

print("Accuracy score of Gradient Boost is:", accuracy_score(y_test, y_pred_gb))
```
