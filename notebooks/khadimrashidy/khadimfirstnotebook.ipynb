**24. Describe in your own words why you use what type of graph for which purpose?**

**Histogram**

Histogram allows us to plot the distribution of numeric variables/valuse

```python
#Example
sns.histplot(df['age'])
```

**Bar Plot**

1.Bar plots help us to visualize the distribution of categorical variables
2.Bar plot shows relation or properties of one distribution variables and categorical variable

```python
sns.barplot(x='smoking_status', y='age', data=df)
```

**Factor Plot**

Factor plots make it easy to separate plots by categorical classes

```python
# Example
sns.factorplot(x='gender', y='heart_disease', data=df)
plt.show()
```

**Density Plot**

Density plots display the ditribution between two variables

```python
# Example
plt.figure(figsize=[10,6])
sns.kdeplot(x='hypertension',y='age', data=df, shade=True)
plt.show()
```

**Joint Distribution Plot**

Joint distribution plots combine information from scatter plots and histograms to give you detailed information for bi-variate distribution

```python
# Example
sns.jointplot(x='age', y='avg_glucose_level', data=df)
plt.show()
```

**25.  Describe the parameters of sns.distplot() in your own words that which parameter is used for what purpose.**

```python
df1=pd.read_csv('../input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv', nrows=400)
df1
```

```python
sns.histplot(data=df1, x='avg_glucose_level')
plt.show()
```

```python
#Flip the plot by assigning the data variable to the y axis:
sns.histplot(data=df1, y='avg_glucose_level')
plt.show()
```

```python
# Check how well the histogram represents the data by specifying a different bin width:
sns.histplot(data=df1, x='age', binwidth=2)
plt.show()
```

```python
# You can also define the total number of bins to use:
sns.histplot(x='age', data=df1, bins=500)
plt.show()
```

```python
# Add a kernel density estimate (kde) to smooth the histogram, providing complementary information about the shape of the distribution:
sns.histplot(x='age', data=df1, kde=True)
plt.show()
```

```python
# If neither x nor y is assigned, the dataset is treated as wide-form, and a histogram is drawn for each numeric column:
sns.histplot(data=df)
plt.show()
```

```python
# We can draw multiple histograms from a long-form dataset with hue mapping:
sns.histplot(x='age', data=df1, hue='work_type' )
plt.show()
```

```python
# The default approach to plotting multiple distributions is to “layer” them, but we can also “stack” them:
plt.figure(figsize=[10,6])
sns.histplot(x='age', data=df1, hue='work_type', multiple='stack')
plt.show()
```

```python
# Overlapping bars can be hard to visually resolve. A different approach would be to draw a step function:
plt.figure(figsize=[10,6])
sns.histplot(x='age', data=df1, hue='work_type', element='step')
plt.show()
```

```python
# We can move even farther away from bars by drawing a polygon with vertices in the center of each bin. 
# This may make it easier to see the shape of the distribution, but use with caution:
# it will be less obvious to your audience that they are looking at a histogram:
plt.figure(figsize=[10,6])
sns.histplot(x='age', data=df1, hue='work_type', element='poly')
plt.show()
```

```python
# To compare the distribution of subsets that differ substantially in size, we can use indepdendent density normalization:
plt.figure(figsize=[10,6])
sns.histplot(x='age', data=df1, hue='work_type', element='step', stat='density',common_norm=False)
plt.show()
```

```python
# It’s also possible to normalize so that each bar’s height shows a probability, which make more sense for discrete variables:
plt.figure(figsize=[10,6])
sns.histplot(x='age', data=df1, hue='work_type', stat='probability', discrete=True)
plt.show()
```

```python
# Real-world data is often skewed. For heavily skewed distributions, it’s better to define the bins in log space. Compare:
# To the log-scale version:
plt.figure(figsize=[10,6])
sns.histplot(x='age', data=df1, log_scale=True, fill=False)
plt.show()
```

```python
# When both x and y are assigned, a bivariate histogram is computed and shown as a heatmap:
plt.figure(figsize=[10,6])
sns.histplot(x='age', y='smoking_status', data=df1, hue='smoking_status', legend=True)
plt.show()
```

```python
# To annotate the colormap, add a colorbar:
plt.figure(figsize=[10,6])
sns.histplot(x='age', y='smoking_status', data=df1, hue='smoking_status', legend=True, cbar=True)
plt.show()
```

### first we try to change any string column in int as in above datafram we can change all string columns in int as;
### For gender column: male= 1 and female=0, Other=2
### For ever_married: Yes=1, No=0
### work_Type: Privite=0, self-employed=1, Gov_job=2, children=3, and Never_worked=4
### Residence_type: Urban=0, Rural=1
### smoking_status: formerly_smoked=0, never_smoked=1, smokes=2, Unknown=3

```python
df1=df.replace(['Male', 'Female',
                    'Yes','No',
                    'Private', 'Self-employed','Govt_job','children','Never_worked','Urban','Rural',
                   'formerly smoked','never smoked','smokes','Unknown'],
                   [1,0,
                    1,0,
                    0,1,2,3,4,
                    0,1,
                    0,1,2,3])
df1
```

```python
y=df1['gender']#output
x=df1.drop(['gender'], axis=1)#input
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.1, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
pred=lr.predict(x_test)
print(pred)
```

```python
print(classification_report(pred,y_test))
```
