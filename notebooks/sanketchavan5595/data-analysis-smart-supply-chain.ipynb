### Which year had the minimum sales for "Nike" products?

```python
df
```

```python
# creating a new column for nike products 
df['nike'] = df['Product Name'].apply(lambda x: x.split(' ')[0] if 'Nike' in x else 0 )
```

```python
nike_df = df[df['nike'] == 'Nike']
```

```python
nike_df
```

```python
nike_df['year'] = nike_df['order date (DateOrders)'].apply(lambda x: x.split('/')[-1].split(' ')[0])
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
plt.figure(figsize = (7,5), dpi = 100)
plt.title('Year v/s Sales for Nike Products')
sns.barplot(x = 'year', 
            y = 'Sales', 
            data = nike_df);
```

```python
nike_df.groupby('year')['Sales'].sum()
```

```python
nike_df.groupby('year')['Sales'].mean()
```
