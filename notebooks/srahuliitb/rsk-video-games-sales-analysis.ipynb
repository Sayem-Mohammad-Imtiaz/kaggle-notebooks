```python
# Total year-on-year sales.
total_yoy_sales = df[['Year', 'NA_Sales','EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].groupby(by = 'Year', as_index = False).sum()
total_yoy_sales
```

```python
total_yoy_sales.columns
```

```python
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.figure(figsize = (12, 5), dpi = 96)
plt.title("Total Year-on-year Sales", fontsize = 12)
for col in total_yoy_sales.columns[1:]:
    plt.plot(total_yoy_sales['Year'], total_yoy_sales[col], label = col)
plt.legend()
plt.show()
```

The most number of games were sold in the year 2008. The number of games solds in a year kept rising until 2008. Afterwards it declined sharply until 2020. Hence, the period of 12 years between 2009 and 2020 suggests that it may not be the best idea for a publisher to invest in console-based games.

```python
# Genre-wise total sales.
genre_wise_total_sales = df[['Genre', 'NA_Sales','EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].groupby(by = 'Genre', as_index = False).sum()
genre_wise_total_sales
```

```python
plt.figure(figsize = (12, 5), dpi = 96)
plt.title("Genre-wise Total Sales", fontsize = 14)
for col in genre_wise_total_sales.columns[1:]:
    plt.plot(genre_wise_total_sales['Genre'], genre_wise_total_sales[col], label = col)
plt.legend()
plt.show()
```

In Japan, "Role-Playing" is the most popular genre. Otherwise, "Action" and "Sports" type games were the best sellers

```python
# Platform-wise total sales.
platform_wise_total_sales = df[['Platform', 'NA_Sales','EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].groupby(by = 'Platform', as_index = False).sum()
platform_wise_total_sales
```

```python
plt.figure(figsize = (12, 5), dpi = 96)
plt.title("Platform-wise Total Sales", fontsize = 14)
for col in platform_wise_total_sales.columns[1:]:
    plt.plot(platform_wise_total_sales['Platform'], platform_wise_total_sales[col], label = col)
plt.xticks(rotation = 90)
plt.legend()
plt.show()
```

The PlayStation2 games were sold the most.

```python
# Publisher-wise total sales.
publisher_wise_total_sales = df[['Publisher', 'NA_Sales','EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].groupby(by = 'Publisher', as_index = False).sum()
# Top 10 best selling publishers globally.
publisher_wise_total_sales.sort_values(by = 'Global_Sales', ascending = False).head(10)
```

```python
# Top 10 best selling publishers in North America.
publisher_wise_total_sales.sort_values(by = 'NA_Sales', ascending = False).head(10)
```

```python
# Top 10 best selling publishers in Europe.
publisher_wise_total_sales.sort_values(by = 'EU_Sales', ascending = False).head(10)
```

```python
# Top 10 best selling publishers in Japan.
publisher_wise_total_sales.sort_values(by = 'JP_Sales', ascending = False).head(10)
```

Japanese publishers dominate the Japan region market in terms of sales.

```python
# Top 10 best selling publishers in other regions.
publisher_wise_total_sales.sort_values(by = 'Other_Sales', ascending = False).head(10)
```

Except in Japan, Electronic Arts, Nintendo, Sony Computer Entertainment and Activision are the most selling publishers.

```python
ptable_glob_sales_year_index_genre_col = pd.pivot_table(df, values = 'Global_Sales', 
                                                        index = ['Year'], columns = ['Genre'], 
                                                        aggfunc = np.sum)
ptable_glob_sales_year_index_genre_col
```

The maximum games of the action genere were sold in the year 2009.

```python
# Year in which the shooter games were sold the most globally.
ptable_glob_sales_year_index_genre_col[ptable_glob_sales_year_index_genre_col['Shooter'] == ptable_glob_sales_year_index_genre_col['Shooter'].max()].index
```

In the year 2011, the shooter games were sold the most globally.

```python
# Year in which the role-playing games were sold the most in Japan.
ptable_jp_sales_year_index_genre_col = pd.pivot_table(df, values = 'JP_Sales', 
                                                      index = ['Year'], columns = ['Genre'],
                                                     aggfunc = np.sum)
ptable_jp_sales_year_index_genre_col
```

```python
ptable_jp_sales_year_index_genre_col[ptable_jp_sales_year_index_genre_col['Role-Playing'] == ptable_jp_sales_year_index_genre_col['Role-Playing'].max()].index
```

In the year 2010, the role-playing games were sold the most in Japan.
