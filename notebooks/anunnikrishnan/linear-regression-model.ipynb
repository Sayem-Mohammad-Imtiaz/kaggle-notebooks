```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=2)
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
#### Y = MX +C
#### brainwght =  m1*x1 + m2*x2 +m3*x3 + C
```

```python
# number of coef is equal to number of feature
# every feature will have a coef/ slope
model.coef_
```

```python
X_train.columns
```

```python
model.intercept_
```

slope of a line with three features / dimensions
Y = m1 * x1 + m2 * x2 + m3 * x3 + intercept(c)

```python
#brain weight = -16.87* gender + -24.5* Age + 0.24* Head Size + 456.62
# head-size = 3000
#age = 2
#gender 2
```

```python
X_test
```

```python
pred = model.predict(X_test)
```

```python
len(pred)
```

```python
X_test.shape
```

```python
print(pred)
```

```python
# load metrics for error calculation
from sklearn.metrics import mean_absolute_error , mean_squared_error
```

```python
mae = mean_absolute_error(Y_test , pred)
print("mae = ",mae)
```

```python
mse = mean_squared_error(Y_test , pred)
print("mse = ",mse)
```

```python
#root_mean_sqr error
rmse = np.sqrt(mean_squared_error(Y_test , pred))
print("rmse = ",rmse)
```

```python
#R2 score - it helps to predict the variations in Y given x1,x2,x3 features
# it does not have ability to determine overfit, it is rarely used in today world
model.score(X_train, Y_train)

# always use rmse, mse, mape , mae to measure your model
```

```python
# mean absolute percentage error
mape1 = np.mean(np.abs((Y_test - pred)/Y_test))*100
```

```python
accuracy = 100- mape1
print("accuracy = ",accuracy)
```
