```python
def find_outliers_tukey(x):
    """
    This function takes Actual-Expected as arg and return outliers indices , values and threshold.
    :param: pandas data frame  object as Difference of Actual - Predicted
    :return value1: indices
    :return value2: outlier values and threshold values upper and lower limit
    :rtype : List. outlier_indices and outlier_values returned as List
     """
    q1 = np.percentile(x, 25)
    q3 = np.percentile(x, 75)
    iqr = q3 - q1
    floor = q1 - 3.0 * iqr
    ceiling = q3 + 3.0 * iqr
    outlier_indices = list(x.index[(x < floor) | (x > ceiling)])
    outlier_values = list(x[outlier_indices])
    return outlier_indices, outlier_values, floor, ceiling


```

```python
### removed outliers
for col in data.columns:
    outlier_indices, outlier_values, floor, ceiling = find_outliers_tukey(data[col])
    #data = data[df[col] != outlier_values]
    data = data[~(data[col] > ceiling) | (data[col] < floor)]
```

```python
# diagnosis is depended variable/ Target
# so by removing diagnosi from data , i get features
X = data.drop('diagnosis',axis =1)
# y is the dependent variable
Y = data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.30, random_state = 2)
```

```python
# pass features and dependent varaible
# fit means model training

for k in range(1,20):
    my_first_model = KNeighborsClassifier(n_neighbors= k)
    my_first_model.fit(X_train, y_train)
    prediction = my_first_model.predict(X_test)
    accuracy = accuracy_score(y_test,prediction)
    print("For k = ",k," the accuracy is: ",accuracy)
```

```python
best_model = KNeighborsClassifier(n_neighbors= 9)
best_model.fit(X_train, y_train)
prediction = best_model.predict(X_test)
accuracy = accuracy_score(y_test,prediction)
print("For k = ",k," the accuracy is: ",accuracy)
```
