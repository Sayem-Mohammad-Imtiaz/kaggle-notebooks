```python
df = pd.read_csv("/kaggle/input/pm25-data-for-five-chinese-cities/BeijingPM20100101_20151231.csv")
df
```

```python
#对美国监测的数据进行绘制
#1、先处理时间：时间戳（DataFrame） or 时间段（PeriodIndex）？
#此处用PeriodIndex对原数据中的时间转换成pandas中的时间序列
period = pd.PeriodIndex(year=df["year"],month=df["month"],day=df["day"],hour=df["hour"],freq="H")
# period
df["date_time"] = period
# df.set_index(["datetime"],inplace=True)
# df
#看一下PM_US Post有多少NAN数据
print("PM_US Post有：",df["PM_US Post"].shape[0],"条数据,其中NaN数据有：",df["PM_US Post"].isna().sum())
# PM_US Post有： 52584 条数据,其中NaN数据有： 2197---->50387条非空
```

```python
# 2、期望用pandas的时间序列方法，必须先把时间序列设置为index
df.set_index(df["date_time"],inplace=True)
# df.head(20)
```

```python
# 3、按月份进行resample（均值）
df_1=df.resample("M").mean()
data = df_1["PM_US Post"]
data_dropna = df_1["PM_US Post"].dropna()

# print(f"data的长度为：{len(data)}，dropna后data的长度为{len(data_dropna)}")
```

```python
# 3.1、绘制data
_x = data.index
_y = data.values
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
plt.figure(figsize=(20,8),dpi=80)
plt.plot(range(len(_x)),_y)
plt.title("PM2.5 data from USA")
plt.show()
#结果太过于稠密，希望变得稀疏一点
```

```python
# 3.2、绘制data_dropna
_x = data_dropna.index
_y = data_dropna.values
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
plt.figure(figsize=(20,8),dpi=80)
plt.plot(range(len(_x)),_y)
plt.title("PM2.5 data from USA(drop_na)")
plt.show()
#结果太过于稠密，希望变得稀疏一点，要去掉na的列数
```

```python
#4、按天数进行resample（均值）
df_2 = df.resample("7D").mean() #按7D降采样（mean）
data_dropna_resample_d = df_2["PM_US Post"].dropna()
```

```python
# 4.1、绘制data_dropna_resample_m
_x = data_dropna_resample_d.index
_y =data_dropna_resample_d.values
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
plt.figure(figsize=(20,8),dpi=80)
plt.plot(range(len(_x)),_y)
plt.title("PM2.5 data from USA resample_7D")
plt.show()
#结果比较满意
```
