## 6. Use the groupby function to display the top 10 publishers with the most global sales with the highest on top.
**Categories:** <br>
(1) .head <br>
(5) .sum <br>
(6) .groupby <br>
(9) .sort_values

```python
video_games.groupby('Publisher')\
    .sum() \
    .Global_Sales \
    .sort_values(ascending=False) \
    .head(10)
    
```

## 7. Create a horizontal bar chart that displays the platform and genre of the top 10 games with the most Japanese Sales. Keep the highest on top.
**Categories:** <br>
(1) .tail <br>
(5) .sum <br>
(6) .groupby <br>
(8) .plot.barh <br>
(9) .sort_values

```python
video_games.groupby(['Platform','Genre']) \
    .JP_Sales \
    .sum() \
    .sort_values() \
    .tail(10) \
    .plot.barh()
```

## 8. Create a pie chart that displays the percentage of all videogames by genre. 
**Categories:** <br>
(7) .groupby().size() <br>
(8) .plot.pie

```python
video_games.groupby('Genre') \
    .size() \
    .plot.pie()
```
