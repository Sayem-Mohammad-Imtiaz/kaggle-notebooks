```python
import pandas as pd
import numpy as np

data = pd.read_csv('../input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')
classify = ["gender","age","hypertension","avg_glucose_level","smoking_status"]
data0 = data[classify]

df = pd.get_dummies(data0["gender"])
df0 = pd.get_dummies(data0["smoking_status"])

gender_column = df.drop(["Male","Other"],axis=1)
smoke_column = df0.drop(["Unknown","formerly smoked","never smoked"],axis=1)

first_version = pd.concat((gender_column,data0),axis=1)
sec = first_version.drop(["gender"],axis=1)
thi = sec.rename(columns={"Female":"Gender"})

sec_version = pd.concat((smoke_column,thi),axis=1)
sec_version0 = sec_version.drop(["smoking_status"],axis=1)
print(sec_version0)

 
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

data_features = ["smokes","Gender","age","hypertension","avg_glucose_level"]
y = sec_version0.age 
X = sec_version0

train_X, val_X, train_y, val_y = train_test_split(X,y,random_state=0)
forest_model = RandomForestRegressor(random_state=1)

forest_model.fit(train_X,train_y)
predict = forest_model.predict(val_X)

prediction_list = []
for age in y:
    prediction_list.append(mean_absolute_error(val_y,predict))
best_pred = min(prediction_list)
index = prediction_list.index(best_pred)
best_fit = y[index]

print(f"Most expected heart attack age is {int(best_fit)}")
```
