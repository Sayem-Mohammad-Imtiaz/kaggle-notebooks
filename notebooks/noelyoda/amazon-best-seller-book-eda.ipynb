```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
plt.figure(figsize = (15,10))
sns.countplot("Genre", data = data)
plt.title("Fiction or Not", fontsize = 15)
plt.show()
```

```python
data["Year"].value_counts()
```

```python
plt.figure(figsize = (15,10))
sns.pointplot(x = "Price", y = "User Rating", data= data, color = "navy", ci = None )
plt.title("Price vs User Rating")
plt.show()
```

```python
plt.figure(figsize = (15,10))
sns.barplot(x = data.Year, y = data.Price.mean(), color = "red", ci = None)
plt.show()
```

```python
a = data.groupby("Year").mean()["Price"]
```

```python
b = data["Year"].value_counts().index
```

```python
plt.figure(figsize = (15,10))
sns.barplot(x = b, y = a, color = "red", ci = None)
plt.title("Prices by years", fontsize = 15)
plt.tick_params(axis = "both", which = "major", labelsize = 12, direction = "out")
plt.xlabel("Years",fontsize = 12)
plt.ylabel("Prices", fontsize = 12)
plt.show()
```

```python
data["User Rating"].value_counts().index
```

```python
sns.jointplot(x = data["User Rating"].value_counts().index, 
              y = data["User Rating"].value_counts(),  kind = "scatter")
plt.xlabel("User Rating")
plt.ylabel("Count")
plt.show()
```

```python
type(data["User Rating"])
```

```python
plt.hist(data["User Rating"], cumulative = True)
plt.ylabel ("Count", fontsize = 13)
plt.xlabel ("User Rating", fontsize = 13)
plt.title("User Rating Histogram", fontsize = 15, color = "blue")
plt.show()
```

```python
author = data["Author"].value_counts()
```

```python
plt.figure(figsize = (20,10))
sns.barplot(x = author.head(50).index, y = author.head(50), palette = "rocket_r")
plt.title("Top 50 Author Counts", fontsize = 15)
plt.xlabel("Authors", fontsize = 12)
plt.ylabel("Count", fontsize = 12)
plt.xticks(rotation = 90)
plt.tick_params(axis = "both", which = "major", labelsize = 14, direction = "out")
plt.show()
```

```python
plt.figure(figsize = (15,10))
sns.heatmap(data.corr(), annot = True, cmap = "YlOrRd")
plt.show()
```

```python
plt.figure(figsize = (15,10))
sns.pointplot(x = "User Rating", y = "Reviews", data = data, 
              color = "blue", linestyles= "-" , markers = "o", ci = None)
plt.xlabel("User Rating", fontsize = 12)
plt.ylabel("Reviews", fontsize = 12)
plt.title ("User Rating vs Reviews", fontsize = 15)
plt.show()
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(figsize = (20,10), ncols = 3)
sns.pointplot(x = "User Rating", y = "Reviews", data = data, 
              color = "blue", linestyles= "-" , markers = "o", ci = None, ax= ax1)
sns.pointplot(x = "Year", y = "User Rating", data = data, color = "red", ci = None, ax = ax2)
sns.pointplot(x = "Year", y = "Reviews", data = data, color = "green", ci = None, ax = ax3)

ax1.set_xlabel("User Rating", fontsize = 12)
ax1.set_ylabel("Reviews", fontsize = 12)
ax2.set_xlabel("Year", fontsize = 12)
ax2.set_ylabel("User Rating", fontsize = 12)
ax3.set_xlabel("Year", fontsize = 12)
ax3.set_ylabel("Reviews", fontsize = 12)
ax1.set_title("Review Count vs User Rating", fontsize = 15)
ax2.set_title("Years and User Rating", fontsize = 15)
ax3.set_title("Years and Review Count", fontsize = 15)
plt.suptitle ("User Ratings w/Different Variables", fontsize = 20)
plt.savefig("Descriptive Analysis Amazon Books")
plt.show()
```
