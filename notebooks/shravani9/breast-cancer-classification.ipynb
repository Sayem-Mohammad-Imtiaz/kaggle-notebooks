```python
import numpy as np 
import pandas as pd 
import seaborn as sns
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cancer = datasets.load_breast_cancer()
```

```python
df = pd.DataFrame(
                  cancer['data'],
                  columns=cancer['feature_names']
                  )
df['target'] = pd.Series(cancer['target'])
df.tail()
```

RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
np.random.seed(42)

#Data 
X = df.drop("target", axis=1)
y = df["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model1 = RandomForestClassifier()
model1.fit(X_train, y_train)
model1.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_train_pred = model1.predict(X_train)
y_test_pred = model1.predict(X_test)

print("Accuracy from RandomForestClassifier on training dataset is - ", accuracy_score(y_train, y_train_pred))
print("Accuracy from RandomForestClassifier on testing dataset is - ", accuracy_score(y_test, y_test_pred))
print('----')
display(confusion_matrix(y_test, y_test_pred))
conf_mat1 = confusion_matrix(y_test, y_test_pred)
```

```python
def plot_conf_mat(conf_mat):
  """
  Plot a confusion matrix using Seaborn's heatmap().
  """
  fig, ax = plt.subplots(figsize=(3,3))
  ax = sns.heatmap(conf_mat,
                    annot=True, # Annotate the boxes with conf_mat info
                    cbar=False)
  plt.xlabel("True label")
  plt.ylabel("Predicted label")
  plt.show()
plot_conf_mat(conf_mat1)
```

LinearSVC

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split
np.random.seed(42)

#Data 
X = df.drop("target", axis=1)
y = df["target"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model2 = LinearSVC(max_iter=10000)
model2.fit(X_train, y_train)
model2.score(X_test, y_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix

y_preds2 = model2.predict(X_test)

print("Accuracy from LinearSVC is - ",accuracy_score(y_test, y_preds2))
display(confusion_matrix(y_test, y_preds2))
conf_mat2 = confusion_matrix(y_test, y_preds2)
```

```python
def plot_conf_mat(conf_mat):
  """
  Plot a confusion matrix using Seaborn's heatmap().
  """
  fig, ax = plt.subplots(figsize=(3,3))
  ax = sns.heatmap(conf_mat,
                    annot=True, # Annotate the boxes with conf_mat info
                    cbar=False)
  plt.xlabel("True label")
  plt.ylabel("Predicted label")
  plt.show()
plot_conf_mat(conf_mat2)
```

DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state=0)
clf.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

#training accuracy 
train_pred = clf.predict(X_train)
print("Accuracy from Decision Tree Classifier on training dataset is - ", accuracy_score(y_train, train_pred))

#testing accuracy 
test_pred = clf.predict(X_test)

print("Accuracy from Decision Tree Classifier on testing dataset is - ", accuracy_score(y_test, test_pred))
```
