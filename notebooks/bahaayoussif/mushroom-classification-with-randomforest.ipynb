```python
def label_encoded(feat):
    le=LabelEncoder()
    le.fit(feat)
    print(feat.name,le.classes_)
    return le.transform(feat)
```

```python
data = pd.read_csv('/kaggle/input/mushroom-classification/mushrooms.csv')
data.isnull().sum()
```

```python
data.head(5)
```

```python
for col in data.columns:
    data[str(col)] = label_encoded(data[str(col)])
```

```python
corr=data.corr()
sns.heatmap(corr)
```

```python
y=data['class']
x=data.drop('class',axis=1)
xtrain,xtest,ytrain,ytest = train_test_split (x,y,test_size=0.2,random_state=42)
```

```python
clf=RandomForestClassifier(n_estimators=200, criterion='gini', min_samples_split=5, min_samples_leaf=2, max_features='auto', bootstrap=True, n_jobs=-1, random_state=42) 
clf.fit(xtrain,ytrain)
pred=clf.predict(xtest)
print(f"The Accuracy of the Random Forest Classifier {accuracy_score(ytest,pred)*100}%")
```
