```python
import pandas as pd
data= pd.read_csv('../input/fetal-health-classification/fetal_health.csv')
data.head(5)
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
y=data['fetal_health']
x=data.drop('fetal_health',axis=1)
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
model_clf=RandomForestClassifier(max_depth=2, random_state=0).fit(xtrain,ytrain)
clf_pred=model_clf.predict(xtest)
print(f"Classifier accuracy Score {round(accuracy_score(clf_pred,ytest)*100,4)}%")

```
