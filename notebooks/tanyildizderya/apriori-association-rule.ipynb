## What is the Association Rule ? 

* Association Rules is one of the very important concepts of machine learning being used in market basket analysis.
* Associatio Rule have 3 different metrics; support, confidence, lift


**1.Support **

This measure gives an idea of how frequent an itemset is in all the transactions. 

![](http://miro.medium.com/max/1400/1*bqdq-z4Ec7Uac3TT3H_1Gg.png)

**2.Confidence**

This measure defines the likeliness of occurrence of consequent on the cart given that the cart already has the antecedents.

![](https://miro.medium.com/max/1400/1*E3mNKHcudWzHySGMvo_vPg.png)

**3.Lift**

Lift controls for the support (frequency) of consequent while calculating the conditional probability of occurrence of {Y} given {X}. 

![](https://miro.medium.com/max/1400/1*Rg429lteTXRKdYgCiHmLVw.png)

```python
data = list(df["products"].apply(lambda x:x.split(',')))
data 
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
```

```python
te = TransactionEncoder()
```

```python
te_data = te.fit(data).transform(data)
```

```python
df = pd.DataFrame(te_data,columns=te.columns_)
df.head()
```

```python
dff = apriori(df,min_support=0.01,use_colnames=True)
dff
```

```python
dff['length'] = dff['itemsets'].apply(lambda x:len(x))
dff
```

```python
dff.describe().T
```

```python
dff[(dff['length']==2) & (dff['support']>=0.05)]
```

```python
dff[(dff['length']==3) & (dff['support']>=0.05)]
```

```python
dff[(dff['length']==4) & (dff['support']>=0.05)]
```
