# Problem
Can we predict whether an AirBnB will have a cleaning fee?

# Dataset formatting/cleaning

```python
df = pd.read_csv("../input/airbnb-price-prediction/train.csv")
df.head(10)
```

```python
filtered = df[["log_price", "accommodates", "bathrooms", "number_of_reviews", "review_scores_rating", "bedrooms", "beds", "cleaning_fee", "cancellation_policy", "bed_type", "review_scores_rating", "room_type"]]
filtered.head()
```

```python
filtered.dtypes
```

```python
df['cleaning_fee'].value_counts()
```

```python
filtered.dropna(inplace=True)
```

```python
filtered['cleaning_fee'] = filtered['cleaning_fee'].map({True: 1, False: 0})
filtered.head()
```

```python
#Categorical Predictors
d = pd.get_dummies(filtered, prefix=['cancellation_policy', 'bed_type', "room_type"], columns=['cancellation_policy', 'bed_type', "room_type"])
d.head()
```

```python
d.corr()
```

# Logistic Regression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
inputDF = d[["cancellation_policy_moderate","cancellation_policy_super_strict_30","cancellation_policy_super_strict_60","cancellation_policy_strict","cancellation_policy_flexible", "bed_type_Airbed", "bed_type_Real Bed", "bed_type_Couch", "bed_type_Pull-out Sofa", "bed_type_Futon"]]
outputDF = d[["cleaning_fee"]].values.ravel()

logisticRegr = LogisticRegression(solver='lbfgs')
x_train, x_test, y_train, y_test = train_test_split(inputDF, outputDF, test_size=0.25, random_state=0)
logisticRegr.fit(inputDF, outputDF)

print(logisticRegr.intercept_)
print(logisticRegr.coef_)
```

```python
y_pred = logisticRegr.predict(x_test)
print("RMSE:\n", np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
(y_pred == y_test).sum()/float(y_test.size)
```
