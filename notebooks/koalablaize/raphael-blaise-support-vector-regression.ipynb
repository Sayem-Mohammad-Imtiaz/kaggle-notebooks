```python
dataset = pd.read_csv('../input/student-grade-prediction/student-mat.csv')
dataset = dataset[['studytime', 'failures', 'Dalc', 'Walc','health','absences','traveltime',
                   'G1', 'G2', 'G3']]
print(dataset)# Chossing attributes that matter or affect the students results try different attributes
```

```python
x = dataset.iloc[:, :-1].values
print(x)
y = dataset.iloc[:, -1].values
y = y.reshape(len(y), 1) 
print(y)
# Reshaping my independent variable (y), into a 2d array as i have to perform feature scaling as the 
#support Vector regression formula does not have any co-efficients like the Multiple linear regression and polynomial Regression i did ealier
# The coefficeints scale my data into the same scale so that is why i did not apply feature scaling in polynomial and multiple regression
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=0) # splitting the dataset into training set and test set before
# feature scaling in-order to avoid leaking of data into the test set
```

```python
# applying feature scaling because my data is not in the same scale in-order to increase the prediction accuracy 
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
x_train = sc_x.fit_transform(x_train)
y_train = sc_y.fit_transform(y_train)
# different scale in the IV so i fit_transform, because it contains 
#different values than on my matrix of features in the x_train
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(x_train, y_train.ravel())
```

```python
y_pred = sc_y.inverse_transform(regressor.predict(sc_x.transform(x_test)))
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)) # cannot visualise a 3-d plot visualising vertically in-order to analyse my predicted scores and real scores
```

```python
# Chaecking the Acurracy of my model
from sklearn.metrics import r2_score
print(r2_score(y_test, y_pred))
```
