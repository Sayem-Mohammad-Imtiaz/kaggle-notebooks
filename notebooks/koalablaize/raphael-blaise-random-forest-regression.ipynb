```python
dataset = pd.read_csv('../input/student-grade-prediction/student-mat.csv')
dataset = dataset[['studytime', 'failures', 'Dalc', 'Walc','health','absences','traveltime',
                   'G1', 'G2', 'G3']] # Chossing attributes that matter or affect the students results try different attributes
print(dataset)
```

```python
# Choosing my matrix of features and my IV 
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# Training my model 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.25, random_state =0)
```

```python
# We need to choose the n_estimators this is basically the number of trees for yout RFR(Random Forest Regression) model, i 
# have included the random state as a parameter to and fitted it into the training set
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2) # the values will have 2 decimal places after the point
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1)) # Can't visualise 3-d plot or rather it is hard to analyse it so a vertical presentation makes it easier for analysis 
```

```python
# Checking the accuracy of my model
from sklearn.metrics import r2_score
print(r2_score(y_test, y_pred))
```
