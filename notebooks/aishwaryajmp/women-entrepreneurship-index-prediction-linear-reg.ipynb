# **Importing Dataset**

```python
df.drop(['Country'],axis=1,inplace =True)
df.head()
```

# **HeatMap for Correlation**

```python
sns.heatmap(df.corr(), annot =True)
```

# Correlation of selected Variables
*There should be good correlation between dependent and independent variables. *

```python
df=df[['Entrepreneurship Index','Women Entrepreneurship Index']]
```

# **HeatMap for correlation of the concidered features**

```python
sns.heatmap(df.corr(), annot =True)
```

# **Splitting Dependent and Independent Features**

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

# **Splitting arrays into Training and Testing Arrays**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.25, random_state=0)
```

# **Training Model**

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X_train, y_train)
```

# **Testing Model**

```python
y_pred = reg.predict(X_test)
y_pred
```

# **R2-Score**

```python
from sklearn.metrics import r2_score
print('r2 score: ', r2_score(y_test, y_pred))
```
