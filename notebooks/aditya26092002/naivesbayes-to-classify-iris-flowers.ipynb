```python
import pandas as pd
import numpy as np
```

# Gaining insights from the data

```python
data=pd.read_csv('../input/iris-flower-dataset/IRIS.csv')
data.head()
data.tail()
data.describe()
```

```python
data.head()
```

```python
data.tail()
```

```python
features=data[['sepal_length','sepal_width','petal_length','petal_width']]
```

```python
target=data['species']
```

```python
target.value_counts()
```

**So there are three classes of flowers**

**Converting Text value to Numerical type**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
target=pd.Series(LabelEncoder().fit_transform(target))
```

```python
target.value_counts()
```

**Now the text has been converted to numerical type
Iris-setosa     =   0
Iris-versicolor =  1
Iris-virginica  =   2**

# Splitting data into train and test data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(features,target,test_size=0.15, random_state=42)
```

# Importing Naives Bayes & Prediction

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf=GaussianNB()
clf.fit(X_train,Y_train)
```

```python
pred=clf.predict(X_test)
print(pred)
```

# Classification Report

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(pred,Y_test))
```
