```python
df = pd.read_csv('../input/iris-flower-dataset/IRIS.csv')

print('DataType in Dataset')
print(df.dtypes)
print('Number of Columns containing Null Value')
print(df.isnull().any().sum(), ' / ', len(df.columns))
print('Number of rows containing null in either column')
print(df.isnull().any(axis=1).sum(), ' / ', len(df))

print('Checking colinearity with species')
#print(df.corr().abs().unstack().sort_values()['species'])


df.head(7)
```

```python
from sklearn import preprocessing # Import label encoder 
label_encoder = preprocessing.LabelEncoder() # label_encoder object knows how to understand word labels.
df['species']= label_encoder.fit_transform(df['species']) # Encode labels in column 'species'. 
df['species'].unique() 
```

```python
print('Checking colinearity with species')
print(df.corr().abs().unstack().sort_values()['species'])
```

```python
 for column in ['sepal_width','sepal_length','petal_length','petal_width']:
    
    sns.boxplot(x=column,y='species',data=df)
    plt.show()

```

```python
#To define x and y
x = df.loc[:, df.columns != 'species'] #to  select multiple column except one data point may be that we want to predict
y=df['species'].values #.values = to get the numpy array and dataset dont return index value and column with selected column

#To Normalise the equation
x=preprocessing.normalize(x)

#print(x.head())
print(y)
```

```python
#logestic regression
x_train, x_test, y_train, y_test = dsplit(x, y, random_state = 1)
reg = LogisticRegression()
reg.fit(x_train, y_train)
predicted = reg.predict(x_test)
from sklearn.metrics import accuracy_score
print(reg.score(x_train, y_train))
print(reg.score(x_test, y_test))
```
