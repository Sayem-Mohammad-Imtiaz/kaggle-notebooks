```python
import numpy as np
import pandas as pd
```

```python
allstar = pd.read_csv('../input/all_star.csv')
allstar.drop(['year', 'game_num', 'game_id', 'team_id', 
              'league_id', 'gp', 'starting_pos'], axis=1, inplace=True)
allstar_count = pd.DataFrame(allstar['player_id'].value_counts())
allstar.set_index('player_id', inplace=True)
allstar = allstar.join(allstar_count)
allstar.reset_index(inplace=True)
allstar.drop_duplicates(inplace=True)
allstar.set_index('index', inplace=True)
allstar
```
