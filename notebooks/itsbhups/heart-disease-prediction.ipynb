```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
dataset = pd.read_csv('/kaggle/input/heart-disease-prediction-using-logistic-regression/framingham.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.isnull().sum()
```

```python
dataset.dropna(inplace = True)
```

```python
dataset.isnull().sum()
```

```python
X = dataset.iloc[:,:-1].values
Y=dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Y_train , Y_test= train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train= sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train) 
print(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 42)
classifier.fit(X_train , Y_train)
Y_pred = classifier.predict(X_test)
from sklearn.metrics import confusion_matrix , accuracy_score
cm = confusion_matrix(Y_pred, Y_test)
print(cm)
accuracy_score(Y_pred, Y_test)
```

```python
print(Y_pred)
print(Y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier_k = KNeighborsClassifier(n_neighbors = 5)
classifier_k.fit(X_train, Y_train)
Y_pred_k =  classifier_k.predict(X_test)
from sklearn.metrics import confusion_matrix , accuracy_score
cm1 = confusion_matrix(Y_pred_k, Y_test)
print(cm1)
accuracy_score(Y_pred_k , Y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier2 =RandomForestClassifier(n_estimators = 200 , random_state =42)
classifier2.fit(X_train , Y_train)
Y_pred3 = classifier2.predict(X_test)
from sklearn.metrics import confusion_matrix , accuracy_score
cm3 =confusion_matrix(Y_pred3, Y_test)
print(cm3)
accuracy_score(Y_pred3, Y_test)
```
