```python
import pandas as pd
import numpy as np
from mlxtend import frequent_patterns
#apriori, association_rules

data = pd.read_csv('/kaggle/input/supermarket/GroceryStoreDataSet.csv', names = ['items'], sep = ',')

```

```python
data.head()
```

```python
data_list = [x.split(",") for x in list(data['items'].values)]
print(data_list[:5])
```

```python
from mlxtend.preprocessing import TransactionEncoder
txn_encode = TransactionEncoder()
predictions = txn_encode.fit(data_list).transform(data_list)
features = pd.DataFrame(predictions,columns=txn_encode.columns_)
features = features.replace(False,0)
features
```

```python
freq_itemsets = frequent_patterns.apriori(features, min_support = 0.2, use_colnames = True, verbose = 1)
freq_itemsets['length'] = freq_itemsets['itemsets'].apply(lambda x: len(x))
freq_itemsets.sort_values(by=['support','length'],ascending=False)
```

```python
# rules selected by lift
association_rules = frequent_patterns.association_rules(freq_itemsets, metric = "lift", min_threshold = 1.2)
association_rules
```

```python
# rules selected by confidence
association_rules = frequent_patterns.association_rules(freq_itemsets, metric = "confidence", min_threshold = 0.5)
association_rules
```
