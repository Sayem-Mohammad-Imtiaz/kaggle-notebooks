```python
def unique_values_of_col(df):
    for column in df : 
        if df[column].dtypes == 'object':
            print(f'{column}: {df[column].unique()}')
```

```python
unique_values_of_col(df)
```

```python
s = (df.dtypes == 'object')
object_cols = list(s[s].index)

print('categorical variables: ',object_cols)
```

```python
from sklearn.preprocessing import OrdinalEncoder

df1 = df.copy()

encoder = OrdinalEncoder()

df1[object_cols] = encoder.fit_transform(df1[object_cols])

df1.head()
```

```python
df1.isnull().sum()
```

```python
df1_salary_mean = df1['salary'].mean()
df1['salary'] = df1['salary'].fillna(df1_salary_mean)
```

```python
scaled_cols = ['ssc_p','hsc_p','degree_p','etest_p','mba_p','salary']

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df1[scaled_cols] = scaler.fit_transform(df1[scaled_cols])


df1.head()
```

```python
features = df1.drop('status',axis = 1)
label = df1['status']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(features,label,test_size = 0.2 , random_state = 1)

print('Shape of x_train: ', x_train.shape)
print('Shape of y_train: ', y_train.shape)
print('Shape of x_train: ',x_train.shape)
print('Shape of x_test: ',x_test.shape)
```

```python
x_train
```

```python
import tensorflow as tf 
from tensorflow import keras 
```

```python
model = tf.keras.Sequential([
    keras.layers.Dense(10,input_shape = (14,) , activation = 'relu'),
    keras.layers.Dense(1,activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'binary_crossentropy',
    metrics = ['accuracy']
)

model.fit(x_train,y_train,epochs = 100)
```

```python
model.evaluate(x_test,y_test)
```

```python
yp = model.predict(x_test)
yp[:5]
```

```python
y_pred = [] 
for element in yp : 
    if element > 0.5 : 
        y_pred.append(1)
    else : 
        y_pred.append(0)
```

```python
y_pred[:5]
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print('Classification Report')
print(classification_report(y_test,y_pred))
```

```python
score = accuracy_score(y_test,y_pred)
print(' Accuracy Score: {:0.2f}'.format(score))
```

```python
cm = confusion_matrix(y_test,y_pred) 

plt.figure(figsize = (5,4))
sns.heatmap(cm , annot = True)

plt.xlabel('Predicted values')
plt.ylabel('Actual values')
plt.title('Confusion Matrix')
plt.show()
```
