```python
df.drop(['victory_status','increment_code', 'white_id', 'black_id', 'moves', 'opening_eco', 'opening_name', 'created_at', 'last_move_at', 'id'], inplace = True, axis = 1)
```

```python
df["rated"]=np.where(df["rated"]==True,1,0)
df["winner"]=np.where(df["winner"]=='white',1,0)
df
```

```python
df.isna().sum()
```

```python
import time
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
```

```python
gnb = GaussianNB()
```

```python
df.columns
```

```python
X_train, X_test = train_test_split(df, test_size=0.5, random_state=int(time.time()))
```

```python
# x_train, x_test, y_train, y_test = train_test_split(df.x, df.y)
```

```python
used_features =[
    "rated",
    "turns",
    "winner",
    "white_rating",
    "black_rating",
    "opening_ply"
]

x_val = X_train[used_features].values
y_val = X_train["winner"]

gnb.fit(x_val, y_val)
```

```python
y_pred = gnb.predict(X_test[used_features]).reshape(-1,1)
```

```python
y_pred
```

```python
# y_val = df[['x']].values.reshape(-1,1)
```

```python
gnb.score(x_val, y_pred)
```
