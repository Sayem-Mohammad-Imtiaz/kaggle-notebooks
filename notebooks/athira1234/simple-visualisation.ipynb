```python
df["percent"]=df["totalscore"]/3
```

```python
def perc(row):
    
        if(row.percent>=90.00):
            val='O'
        elif(row.percent>=80.00 and row.percent<90.00):
            val='A'

        elif(row.percent>=70.00 and row.percent<80.00) :
            val='B'
          
        elif(row.percent>=60.00 and row.percent<70.00):
            val='C'
           
        elif(row.percent>=45.00 and row.percent<60.00):
            val='D'
           
        else:
            val='F'
        return val    


            
```

```python
df['Grade']=df.apply(perc,axis=1)
```

```python
pd.crosstab(df.Grade,df.gender)
```

Female students have scored more Os
than male students
