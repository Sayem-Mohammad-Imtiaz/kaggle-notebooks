# The kaggle working week
Just for fun I thought I would take a look at the activity on kaggle for each day of the week, since records began...

```python
import pandas as pd
import plotly.express as px
diasDeLaSemana = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
# read in the data
competitions_df = pd.read_csv('../input/meta-kaggle/Submissions.csv')
# add a column of 1's to sum up
competitions_df['ones'] = 1
# convert the date/times
competitions_df['SubmissionDate'] = pd.to_datetime(competitions_df['SubmissionDate'])
# sum things up by the day of the week
week_df = competitions_df.groupby(competitions_df['SubmissionDate'].dt.day_name()).sum().reindex(diasDeLaSemana)
```

```python
# and now make a plot
px.bar(week_df, y=['ones'], color=week_df['ones'].values , title="Competition submissions").show()
```

```python
# read in the data
datasets_df = pd.read_csv('../input/meta-kaggle/Datasets.csv')
# add a column to sum up
datasets_df['ones'] = 1
# convert the date/times
datasets_df['CreationDate'] = pd.to_datetime(datasets_df['CreationDate'])
# sum things up by the day of the week
week_df = datasets_df.groupby(datasets_df['CreationDate'].dt.day_name()).sum().reindex(diasDeLaSemana)
```

```python
px.bar(week_df, y=['ones'], color=week_df['ones'].values , title="Dataset creation").show()
```

# Discussion
* **Competitions:** We can see that the busiest days are Monday and Tuesday, both having approximately 20% more activity compared to Saturday. We can see that Sunday is slightly more active than Saturday, this may be due to people getting in last minute submissions ready for the traditional Monday competition deadline (or Tuesday morning, depending on your time zone).
* **Datasets:** Wednesday is the most popular day for submitting datasets to kaggle, with approximately 30% more activity compared to Saturday.
* **Notebook creation:** Tuesday is the most popular day for creating notebooks, again with approximately 30% more activity compared to the weekend.
* **Topics:** The most active days for Topic creation are Tuesday, Wednesday and Thursday, with approximately 30% more activity compared to the weekend. The Discussion replies chart, unsurprisingly, closely parallels the Topic posts, with 50% more activity with respect to the weekend. The 'replies' chart is more or less 6 times taller than the Topic chart, indicating that on average each Topic receives around six replies.
