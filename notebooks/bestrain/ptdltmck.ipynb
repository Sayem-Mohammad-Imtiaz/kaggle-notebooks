# Số ca nhiễm theo thời gian của toàn thế giới:

# Tỷ lệ tử vong (death/confirmed)

```python
temp_df=pd.DataFrame(death_df.groupby(['Country/Region']).sum()[[death_df.columns[-1]]]/conf_df.groupby(['Country/Region']).sum()[[conf_df.columns[-1]]]).rename(columns={"8/12/20": "death rate"})
temp_df['recovered rate']=pd.DataFrame(recov_df.groupby(['Country/Region']).sum()[[recov_df.columns[-1]]]/conf_df.groupby(['Country/Region']).sum()[[conf_df.columns[-1]]])
```

```python
temp_df.sort_values(by=['death rate'],ascending=False)[:10].plot.barh(y=['death rate'],title='Death rate ( death/confirmed )')
```

<font size="5">  
    
==> Nhận xét:
- Top 10 nước có tỉ lệ tử vong cao nhất
- Top 3 bao gồm Anh, MS Zaandam, Yemen, với Yemen có tỉ lệ tử vong cao nhất khoảng 30%, đây là tỉ lệ khá cao
- Với Ý, Anh, Mexico, Pháp cũng có mặt tại top 10 số cases tử vong, thì cũng xuất hiện trong top 10 tỉ lệ tử vong cao nhất. Với Anh số ca tử vong cao và tỉ lệ tử vong cũng đều cao so với các nước khác
    
</font>

```python
temp_df.sort_values(by=['recovered rate'],ascending=False)[:10].plot.barh(y=['recovered rate'],title='Recovered rate ( recovered/confirmed )')
```

<font size="5">  
    
==> Nhận xét:
- Top 10 nước có tỉ lệ bệnh nhân phục hồi cao nhất
- Với top 4 nước đầu có tỉ lệ bệnh nhân đã phục hồi là 100%, tức là tất cả đều đã khỏi bệnh, không có ca tử vong
- Malaysia có tỉ lệ phục hồi cũng khá cao, khoảng 95%
- Các nước này không có có số case nhiễm không nằm trong top các nước có số cases nhiễm và số tử vong cao nhất
- Có lẽ việc không để dịch bùng phát mạnh là việc quan trọng nhất, với số ca nhiễm khá cao thì khó mà kiểm soát được tình hình, dễ gây quá tải và việc giảm ca tử vong cũng như việc điều trị cho bênh nhân mới có kết quả tốt.c

## Sự lây lan trên toàn thế giới - GeoMap

### Số cases nhiễm toàn cầu - GeoMap

```python
geo_df=df.groupby(['ObservationDate', 'Country/Region'])[['Confirmed', 'Deaths']].sum().reset_index()
geo_df.head()
```

```python
fig = px.scatter_geo(geo_df, locations="Country/Region", 
                    locationmode='country names', color="Confirmed", 
                    hover_name="Country/Region", range_color=[0,geo_df.max()['Confirmed']], 
                     animation_frame='ObservationDate',
                    color_continuous_scale="Bluered", 
                      projection="natural earth",size='Confirmed',
                    title='Countries with Confirmed Cases')
fig.show()
```

<font size="5">  
    
==> Nhận xét:
- Dịch covid bùng phát ở Trung quốc và một số nước Châu Á như Hàn Quốc, Philippines, Malaysia, Nhật Bản khoảng từ đầu dịch tới tháng 3 thì Trung Quốc và Hàn Quốc là 2 ổ dịch lớn của thế giới
- Đầu tháng 3 dịch đã bùng phát đến các Nước Ở Châu Âu, và Ý là ổ dịch lớn thứ 3 sau Trung Quốc Và Hàn Quốc
- Cũng vào đầu tháng 3, Số case nhiễm ở Trung quốc chững lại, dịch có vẻ được kiểm soát, với số cases nhiễm dừng lại ở khoảng 80k
