# **India from 2006 to 2020**

```python
India=df[df['Country name']=='India']
```

```python
India
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=India,x="year",y="Life Ladder",marker='o')
plt.title("Life Ladder")
plt.show()
```

**There are certain peaks with regards to India's ladder score as seen in the years 2008, 2010, 2012 and 2020. There are also certain drops as seen in the years 2007, 2009, 2011 and an overall low in 2019. A continouse decrease can be seen from 2012 to 2019. The overall trend for the ladder score is decreasing.**

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=India,x="year",y="Log GDP per capita",marker='o')
plt.title("Log GDP per capita")
plt.show()
```

**India's GDP is on an increasing trend over the years, with the exception of 2020.**

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=India,x="year",y="Social support",marker='o')
plt.title("Social support")
plt.show()
```

**There are some distinct lows in 2007, 2012 and 2019, 2012 being the lowest and some distinct highs in 2008, 2014 and 2018. India's social support is on a decreasing trend over the years.**

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=India,x="year",y="Healthy life expectancy at birth",marker='o')
plt.title("Healthy life expectancy at birth")
plt.show()
```

**An increasing trend over the years.**

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=India,x="year",y="Freedom to make life choices",marker='o')
plt.title("Freedom to make life choices")
plt.show()
```

**There's a slight increasing trend, but there are distincts drops as seen in years 2008, 2012, 2015 and 2019. With an all time low in year 2012.There are certain peaks with regards to India's Freedom to make life choices as seen in the years 2011, 2014, 2017 and 2020.**

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=India,x="year",y="Generosity",marker='o')
plt.title("Generosity")
plt.show()
```

**Generosity amongst Indians have increased over time, with spikes in 2010, 2013, 2016 and 2019 and an all time low in 2008**

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=India,x="year",y="Perceptions of corruption",marker='o')
plt.title("Perceptions of corruption")
plt.show()
```

**There were peaks in 2011, 2014 and 2018, and the perception of corruption has decreased over the years.**

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=India,x="year",y="Positive affect",marker='o')
plt.title("Positive affect")
plt.show()
```

**There were peaks in 2009, 2014 and 2020**

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=India,x="year",y="Negative affect",marker='o')
plt.title("Negative affect")
plt.show()
```

**An increasing trend over the years.**

```python
df2 = world.set_index('Country name')
temp = pd.DataFrame(df2['Ladder score']).reset_index()

#ADAPTING TO THE ISO 3166 STANDARD
temp.loc[temp['Country name'] == 'Taiwan Province of China', 'Country name'] = 'Taiwan, Province of China' 
temp.loc[temp['Country name'] == 'Hong Kong S.A.R. of China', 'Country name'] = 'Hong Kong' 
temp.loc[temp['Country name'] == 'Congo (Brazzaville)','Country name'] = 'Congo' 
temp.loc[temp['Country name'] == 'Palestinian Territories','Country name'] = 'Palestine, State of' 

temp.drop(index=temp[temp['Country name'] == 'Kosovo'].index, inplace=True) 
temp.drop(index=temp[temp['Country name'] == 'North Cyprus'].index, inplace=True) 


temp['iso_alpha'] = temp['Country name'].apply(lambda x:pc.country_name_to_country_alpha3(x,))
fig = px.choropleth(temp, locations='iso_alpha',
                    color='Ladder score',
                    hover_name='Country name',
                    color_continuous_scale=px.colors.diverging.RdYlGn,
                   )
fig.update_layout(
    title_text='World map - Ladder score',
    showlegend=False,
    paper_bgcolor='rgb(248, 248, 255)',
    geo_bgcolor='rgb(248, 248, 255)',
    geo_showframe=False,
    title_font_size=22,
)
fig.show()
```
