### Lets load the data into a pandas dataframe and view its basic properties

```python
df = pd.read_csv("/kaggle/input/salary-data-simple-linear-regression/Salary_Data.csv")

df.describe()
```

### Now we will separate predictor dataset from target data set
these two will be used to train and test our model

```python
X = df[["YearsExperience"]] # this is is the predictor variable
y = df[["Salary"]] # this is the target variable
```

### Let us split the training data and testing data which we will use on the model

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

### Time to fit the training data in model and test it against test data to get the R2 Score

```python
from sklearn.linear_model import LinearRegression

model =LinearRegression()

model.fit(X_train,y_train)

y_predict = model.predict(X_test)


from sklearn.metrics import r2_score
print("R-square metric : ", r2_score(y_test,y_predict))
```

## The above R-Square metric value shows that for a given predictor variable value, the model is able to predict the value of target variable with 98% accuracy
## this is a good score for our model!
