# This notebook is to perform simple linear regression usning Boston housing dataset 
### we will consider the column, ‘RM’, as predictor, and ‘MEDV’ as the target variable

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

df_housing = pd.read_csv("/kaggle/input/boston-housing-dataset/HousingData.csv") 

df_housing.describe()
```

### Consider the column, ‘RM’, as predictor, and ‘MEDV’ as the target variable

```python
df = df_housing[["RM","MEDV"]]

predictor = df_housing[["RM"]]
target = df_housing[["MEDV"]]
```

###  Visualize the association between the predictor and the target using scatter plot.

```python
import matplotlib.pyplot as plt

plt.scatter(predictor,target)
```

### Split the data into train and test datasets, in the ratio of 67:33

```python
from sklearn.model_selection import train_test_split

train_pd,test_pd,train_tr,test_tr = train_test_split(predictor,target,test_size=0.33,random_state=42)
```

### Build a Linear Regression model using training dataset, to predict the target variable.

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(train_pd,train_tr)

predicted = model.predict(test_pd)


print("Intercept:",model.intercept_)
print("Coefficients:",model.coef_)
print("Coeff of Determination",model.score(test_pd, test_tr))
```

```python
from sklearn.metrics import r2_score

print("r2 score : ",r2_score(test_tr,predicted))

from sklearn.metrics import mean_squared_error

print("MSE : ",mean_squared_error(test_tr,predicted))
```

## R Square/Adjusted R Square is better used to explain the model to other people because you can explain the number as a percentage of the output variability. MSE, RMSE, or MAE are better be used to compare performance between different regression models.
