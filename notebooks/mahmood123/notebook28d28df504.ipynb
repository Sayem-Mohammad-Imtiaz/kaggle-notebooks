```python
data = pd.read_csv("/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv")
data.head()
```

```python
data["iso_code"]= data.iso_code.replace(np.nan, "Unknown");data

data = data.replace(np.nan, 0);
```

The question I will address is “What are the most popular and preferred vaccines around the world?”

Finding out the most popular vaccines/combination of vaccines that is used by different countries

```python
pop_vaccines = data.groupby(["vaccines"])['total_vaccinations'].max()

pop_vaccines.sort_values(ascending = False).plot.bar(color = "red", xlabel = "Vaccines", ylabel = "NO of people vaccinated(in crores)"

                                                     , title = "Most popular combinations vaccines", figsize = (12,6))
```

This bar plot shows the most popular combinations of vaccines in the world. They are Johnson&Johnson, Moderna, and Phizer/Biontech. This is likely due to their effectivness, their low mortality rates, and production rates. 

```python
data.groupby(["vaccines"])["country"].nunique().sort_values(ascending = False).plot.pie(startangle = 30,fontsize = 10, figsize = (8,10), autopct = "%.1f")
plt.title('Most prefered combination of vaccines')
```

The most preffered combination of vaccines in the world is  Astro Zenica and Oxford.

We can see that the most popular vaccines are Johnson&Johnson, Moderna, and Phizer/Biontech, and the most preffered combination of vaccines are Astro Zenica and Oxford
