 The most interesting target features are rotor temperature ("pm"), stator temperatures ("stator_*") and torque.
Especially rotor temperature and torque are not reliably and economically measurable in a commercial vehicle.

Being able to have strong estimators for the rotor temperature helps the automotive industry to manufacture motors with less material and enables control strategies to utilize the motor to its maximum capability.
> A precise torque estimate leads to more accurate and adequate control of the motor, reducing power losses and eventually heat build-up.

```python
df1=maindf
```

```python
target=df1.pop("torque")

X_train, X_test, y_train, y_test = train_test_split(
    df1, target, test_size=0.33, random_state=42)
```

```python
target
```

# Regression

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
model=reg.fit(X_train,y_train)
pred_y=model.predict(X_test)
print(mean_squared_error(y_test,pred_y))
print(mean_absolute_error(y_test,pred_y))
print(r2_score(y_test,pred_y))
```

```python
clf = make_pipeline(StandardScaler(),
                    SGDRegressor(max_iter=1000, tol=1e-3,penalty='l2'))
model2=clf.fit(X_train, y_train)
pred_y2=model2.predict(X_test)
print(mean_squared_error(y_test,pred_y2))
print(mean_absolute_error(y_test,pred_y2))
print(r2_score(y_test,pred_y2))
```

```python

clf3 = RANSACRegressor(LinearRegression(),random_state=22)
model3=clf3.fit(X_train, y_train)
pred_y3=model3.predict(X_test)
print(mean_squared_error(y_test,pred_y3))
print(mean_absolute_error(y_test,pred_y3))
print(r2_score(y_test,pred_y3))
```
