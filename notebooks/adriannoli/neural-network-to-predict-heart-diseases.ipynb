Now it's necessary divide the data in a vector called "y" and in a matrix called "X"

```python
df = df.sample(frac=1)
y = df.target
X = df.drop("target",axis=1)
```

"y" is the values that we want to predict

"X" is the data to predict "y"

```python
X, y
```

Now we normalized the "X" data with MinMaxScaler() 

```python
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
X
```

It's time to create the training and test data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=100)
print("X_train", X_train.shape)
print ("------------------------------------")
print("X_test", X_test.shape)
print ("------------------------------------")
print("Y_train", y_train.shape)
print ("------------------------------------")
print("Y_test", y_test.shape)
```

Now it's possible program an artificial neural network using Keras.

The NN has the following specifications

* epoch=1000

* batch size=10

* number of neuron in each one layer=10

* 5 layers, the first using softplus, trhee more using softmax and the last with hyperbolic tangent Activation Function

* The optimizer is of Gradient descent (with momentum) stochastic, with a learning rate of 10%

```python
epoch=1000
batch=10
neuronas=10
a1="softplus"
a2="softmax"
model = keras.Sequential([
    layers.Dense(neuronas, activation=a1,input_shape=(13,)),
    layers.Dense(neuronas, activation=a2),
    layers.Dense(neuronas, activation=a2),
    layers.Dense(neuronas, activation=a2),
    layers.Dense(1, activation="tanh")])
sgd = SGD(lr=0.1)
model.compile(optimizer=sgd, loss='mean_squared_error')

model.fit(X_train, y_train, epochs=epoch, batch_size=batch)
```

Now we can predict "y" values

```python
y_hat_train = model.predict(X_train).round() # for traning data
cm_train = confusion_matrix(y_train,y_hat_train)
labels = np.unique(y_hat_train)
sns.heatmap(cm_train, annot=True, fmt="d", xticklabels=labels, yticklabels=labels, cmap="inferno")
plt.ylabel("Predicted")
plt.xlabel("Truth")
plt.title("Confusion Matrix for prediction using training data");
```

```python
print(accuracy_score(y_train,y_hat_train))
print(classification_report(y_train,y_hat_train))
```

```python
y_hat_test = model.predict(X_test).round() # for test data
cm_test = confusion_matrix(y_test,y_hat_test)
labels = np.unique(y_hat_test)
sns.heatmap(cm_test, annot=True, fmt="d", xticklabels=labels, yticklabels=labels, cmap="inferno")
plt.ylabel("Predicted")
plt.xlabel("Truth")
plt.title("Confusion Matrix for prediction using test data");
```

```python
print(accuracy_score(y_test,y_hat_test))
print(classification_report(y_test,y_hat_test))
```
