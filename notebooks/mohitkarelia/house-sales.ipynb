# **Analysis**

```python
plt.figure(figsize = (25,10))
sb.heatmap(data.corr(),annot=True)
```

```python
X = data.iloc[:,1:]
y = data['price']
```

# **Regular Machine Learning**

Without Pre-Processing and without PCA

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import mean_squared_error,r2_score
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge,ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,BaggingRegressor,GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.20,random_state = 1)
```

Linear Regression

```python
model1 = LinearRegression()
model1.fit(X_train,y_train)
```

```python
print("Insample Score:",model1.score(X_train,y_train))
print("Outsample Score:",model1.score(X_test,y_test))
```

```python
y_pred_train = model1.predict(X_train)
y_pred_test = model1.predict(X_test)
```

```python
plt.figure(figsize = (25,10))
sb.kdeplot(y_train,color = 'red',label = "Training value");
sb.kdeplot(y_pred_train,color = 'green',label = "Predicted value");
plt.legend();
plt.title("Insample Comparision");
```

```python
plt.figure(figsize = (25,10))
sb.kdeplot(y_test,color = 'red',label = "Training value");
sb.kdeplot(y_pred_test,color = 'green',label = "Predicted value");
plt.legend();
plt.title("Outsample Comparision");
```

```python
plt.figure(figsize = (30,9))
plt.scatter(y_pred_train,y_pred_train-y_train,color = 'blueviolet',label = 'train_data');
plt.scatter(y_pred_test,y_pred_test-y_test,color = 'red',label = 'test_data');
plt.legend();
plt.title('Residual Analysis');
```

```python
print('mse in sample:', mean_squared_error(y_train,y_pred_train))
print('mse out sample:', mean_squared_error(y_test,y_pred_test))
print('r2score in sample:', r2_score(y_train,y_pred_train))
print('r2score out sample:',r2_score(y_test,y_pred_test))
```
