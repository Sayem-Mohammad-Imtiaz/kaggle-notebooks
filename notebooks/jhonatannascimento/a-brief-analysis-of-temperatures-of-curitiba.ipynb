### Import of the libraries for analysis:

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

### Data processing:

```python
ctba_temp = pd.read_csv("../input/temperature-timeseries-for-some-brazilian-cities/station_curitiba.csv")
```

```python
ctba_temp.head(10)
```

### Data cleaning:

How we can see, the dataset have values divergents of the means of temperatures possibles. That way, we need clean this values for better analysis.

```python
ctba_temp.replace(999.90, np.nan, inplace=True)
ctba_temp.head(10)
```

Other change necessary is to insert 'YEAR' as index, that's will facilit manipulations with dataset:

```python
df_ctba_temp = ctba_temp.set_index("YEAR", inplace=False)
df_ctba_temp
```

### Data analysis

For first, we will analyse the relation between year and mean temperatures. For that, we need define a dataset that express this relation.

```python
mean_temperature = ['metANN']
ctba_mean_temperature = df_ctba_temp[mean_temperature]
```

```python
ctba_mean_temperature.head(10)
```

We can plot a grafic for better visualization:

```python
plt.figure(figsize=(12,6))
sns.lineplot(data = ctba_mean_temperature, legend=False)
plt.ylabel("Mean temperature (°C)")
```

How we can see, the mean temperature present growth over the time, but is not a big variation. 

Continuing our analisys, we can see the mean temperature per months of the 2018, because 2019 have incomplet data. For that, it's necessary create other dataset for to reach this objective.

```python
ctba_temp_month = df_ctba_temp.loc[:, :"DEC"]
```

```python
ctba_temp_2018 = ctba_temp_month[ctba_temp_month.index == 2018]
ctba_temp_2018
```

```python
ctba_temp_2018 = ctba_temp_2018.T
ctba_temp_2018
```

Ploting this datas in a grafic for analysis

```python
plt.figure(figsize=(12, 6))
sns.lineplot(data=ctba_temp_2018, legend=False)
plt.ylabel("Temperature (°C)")
plt.title("Curitiba temperature mean in 2015")
```

The minors mean temperatures are in June and August, these months are winter in Curitiba, that way usually costum be cold in this a epoch. As also, the biggers mean temperatures , December and March, are months of the summer in this region.
