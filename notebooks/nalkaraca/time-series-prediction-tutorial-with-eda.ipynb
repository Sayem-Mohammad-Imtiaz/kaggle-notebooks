```python
# ATTACK  saldırı yapan ulkeler 
# map plot 1
aerial["color"] = ""
aerial.color[aerial.Country == "USA"] = "rgb(0,116,217)"# ulkelerın renklerı belırlenır
aerial.color[aerial.Country == "GREAT BRITAIN"] = "rgb(255,65,54)"
aerial.color[aerial.Country == "NEW ZEALAND"] = "rgb(133,20,75)"
aerial.color[aerial.Country == "SOUTH AFRICA"] = "rgb(255,133,27)"

data = [dict(
    type='scattergeo', # cografya harıtası 
    lon = aerial['Takeoff Longitude'],
    lat = aerial['Takeoff Latitude'],
    hoverinfo = 'text',# noktanın yanına gelınce hangi ulke ıste ısım gösterır text içine yaılır
    text = "Country: " + aerial.Country + " Takeoff Location: "+aerial["Takeoff Location"]+" Takeoff Base: " + aerial['Takeoff Base'],
    mode = 'markers',
    marker=dict(
        sizemode = 'area',
        sizeref = 1,
        size= 10 ,
        line = dict(width=1,color = "white"), # noktanın etrafındakı renk belırler
        color = aerial["color"], # 
        opacity = 0.7),)]# saydamlık belırler

layout = dict(                            # data içinde yaratılır baslık ulkeler kıyışerıtlerı geo özellıklerı
    title = 'Countries Take Off Bases ',
    hovermode='closest',
    geo = dict(showframe=False, showland=True, showcoastlines=True, showcountries=True,
               countrywidth=1, projection=dict(type='Mercator'),
              landcolor = 'rgb(217, 217, 217)',
              subunitwidth=1,
              showlakes = True,        # gollerı gösterır
              lakecolor = 'rgb(255, 255, 255)',
              countrycolor="rgb(5, 5, 5)")    # ulke renklerını gösterır
fig = go.Figure(data=data, layout=layout)    # bırlestırır 
iplot(fig)# gösterır
```

```python
# Bombing paths map plot 2
# trace1
airports = [ dict(                   # kalkıs yerı enlem boylam ulke noktası
        type = 'scattergeo',
        lon = aerial['Takeoff Longitude'],
        lat = aerial['Takeoff Latitude'],
        hoverinfo = 'text',
        text = "Country: " + aerial.Country + " Takeoff Location: "+aerial["Takeoff Location"]+" Takeoff Base: " + aerial['Takeoff Base'],
        mode = 'markers',
        marker = dict( 
            size=5, 
            color = aerial["color"],
            line = dict(
                width=1,
                color = "white")))]

# trace2
targets = [ dict(                     # hedef yerı enlem boylam ulke noktası
        type = 'scattergeo',
        lon = aerial['Target Longitude'],
        lat = aerial['Target Latitude'],
        hoverinfo = 'text',
        text = "Target Country: "+aerial["Target Country"]+" Target City: "+aerial["Target City"],
        mode = 'markers',
        marker = dict( 
            size=1, 
            color = "red",
            line = dict(
                width=0.5,
                color = "red")))]
        
# trace3
flight_paths = []  # boş lıste atanıyor
for i in range( len( aerial['Target Longitude'] ) ): # uzunluk belırlenır
    flight_paths.append(                            # liste içini doldurulur
        dict(type = 'scattergeo',
            lon = [ aerial.iloc[i,9], aerial.iloc[i,16] ],# saldırı yerı kordınatı
            lat = [ aerial.iloc[i,8], aerial.iloc[i,15] ],# kalkıs yerı kordınatı
            mode = 'lines',
            line = dict(
            width = 0.7,
            color = 'black',),
            opacity = 0.6,) )
    
layout = dict(
    title = 'Bombing Paths from Attacker Country to Target ',
    hovermode='closest',
    geo = dict(showframe=False, showland=True, showcoastlines=True, showcountries=True,
               countrywidth=1, projection=dict(type='Mercator'), # harıtada kı yere dık bakmak veya zumlamak gıbı 
              landcolor = 'rgb(217, 217, 217)',
              subunitwidth=1,
              showlakes = True,
              lakecolor = 'rgb(255, 255, 255)',
              countrycolor="rgb(5, 5, 5)")
)
    
fig = dict( data=flight_paths + airports+targets, layout=layout )
iplot( fig )
```
