```python
# prepare data frame
# Line plot

df = timesData.iloc[:100,:] # 100 e kadar yazdır

# import graph objects as "go"
import plotly.graph_objs as go

# Creating trace1
trace1 = go.Scatter(x = df.world_rank,   # kalıp seklinde parametredir
                    y = df.citations,
                    mode = "lines",
                    name = "citations",
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),# renklerı ve saydamlık belırler
                    text= df.university_name)# grafigin üstüne gelınce ıstenılen yazı yazılır
# Creating trace2

trace2 = go.Scatter(x = df.world_rank,
                    y = df.teaching,
                    mode = "lines+markers",
                    name = "teaching",
                    marker = dict(color = 'rgba(80, 26, 80, 0.8)'),# renklerı ve saydamlık belırler
                    text= df.university_name)
data = [trace1, trace2] # burda yapılan kalıpları data ıle birlestırılır
layout = dict(title = 'Citation and Teaching vs World Rank of Top 100 Universities',
              xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False))
fig = dict(data = data, layout = layout) # burda hepsı birlestırilerek çizdirilir
iplot(fig)
```

```python
#Bar plot
# prepare data frames

df2014 = timesData[timesData.year == 2014].iloc[:3,:]# flitreleme yapılır ılk 3 unıversıte alınır.
# import graph objects as "go"

import plotly.graph_objs as go

# create trace1 # bar cizimi kullanılır

trace1 = go.Bar(x = df2014.university_name,
                y = df2014.citations,
                name = "citations",
                marker = dict(color = 'rgba(255, 174, 255, 0.5)',
                             line=dict(color='rgb(0,0,0)',width=1.5)),# çevre kalınlıgı ve rengi 0 oldugundan siyah olur
                text = df2014.country)

# create trace2 # bar cizimi kullanılır

trace2 = go.Bar(x = df2014.university_name,
                y = df2014.teaching,
                name = "teaching",
                marker = dict(color = 'rgba(255, 255, 128, 0.5)',
                line=dict(color='rgb(0,0,0)',width=1.5)),# çevre kalınlıgı ve rengi 0 oldugundan siyah olur
                text = df2014.country)

data = [trace1, trace2]# yapılan kalıpları data ıle birlestırılır

layout = go.Layout(barmode = "group") # grupla anlamına gelır
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

```python
# pie plot
# data preparation
df2016 = timesData[timesData.year == 2016].iloc[:7,:] # 7 inci sıraya kadar alınır
pie1 = df2016.num_students
pie1_list = [float(each.replace(',', '.')) for each in df2016.num_students]  # str(2,4) => str(2.4) = > float(2.4) = 2.4
labels = df2016.university_name
# figure tek bir fıh içinde datalar ıle olusturur
fig = {"data": 
       
[{"values": pie1_list,
 "labels": labels,
 "domain": {"x": [0, .5]},
 "name": "Number Of Students Rates",
 "hoverinfo":"label+percent+name",
 "hole": .3,
 "type": "pie" },],# pie ıle olusturulur
   
  "layout": {
        "title":"Universities Number of Students rates",
        "annotations": [
            { "font": { "size": 20},
              "showarrow": False,
              "text": "Number of Students",
                "x": 0.20,
                "y": 1 },]}}
iplot(fig) # hepsını ekranda görsellestırır 
```

```python
# Bubble plot
# data preparation
df2016 = timesData[timesData.year == 2016].iloc[:20,:]# 20 unıversıte flıtre ederız 
num_students_size  = [float(each.replace(',', '.')) for each in df2016.num_students]# virgul notkaya cevrılır float yapılır
international_color = [float(each) for each in df2016.international]

# farklı kod yontemlerı ıle data yazılır
data = [
    {
        'y': df2016.teaching,
        'x': df2016.world_rank,
        'mode': 'markers',
        'marker': {
            'color': international_color,
            'size': num_students_size,
            'showscale': True},
        "text" :  df2016.university_name}]
iplot(data) # fıg kullanmadan datalar birlestırıler gorsellestırılır.
```

```python
# histogram datanın içinde A degerinin kactane oldugunu söyler
# prepare data
x2011 = timesData.student_staff_ratio[timesData.year == 2011]
x2012 = timesData.student_staff_ratio[timesData.year == 2012]

trace1 = go.Histogram(
    x=x2011,
    opacity=0.75,
    name = "2011",
    marker=dict(color='rgba(171, 50, 96, 0.6)'))
trace2 = go.Histogram(
    x=x2012,
    opacity=0.75,
    name = "2012",
    marker=dict(color='rgba(12, 50, 196, 0.6)'))

data = [trace1, trace2]
layout = go.Layout(barmode='overlay',                              # iç içe gecmelı grafık gösterır barmode kodu
                   title=' students-staff ratio in 2011 and 2012',
                   xaxis=dict(title='students-staff ratio'),
                   yaxis=dict( title='Count'),
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
# scatter plot matrix 
# import figure factory
import plotly.figure_factory as ff
# prepare data

dataframe = timesData[timesData.year == 2015] # 2015 fılıtrelerız
data2015 = dataframe.loc[:,["research","international", "total_score"]]# 3 tane baslık belırlenır 
data2015["index"] = np.arange(1,len(data2015)+1)

# scatter matrix   burda bu kod ıle 2 tane fazla gorsel elde edılır 

fig = ff.create_scatterplotmatrix(data2015, diag='box', index='index',colormap='Portland',
    colormap_type='cat',
    height=700, width=700)# yukseklık ve genıslık

iplot(fig)
```

```python
# ınset plot 
# first line plot
trace1 =go.Scatter(
    x=dataframe.world_rank,
    y=dataframe.teaching,
    name = "teaching",
    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),)
# second line plot
trace2 =go.Scatter(
    x=dataframe.world_rank,
    y=dataframe.income,
    xaxis='x2',
    yaxis='y2',
    name = "income",
    marker = dict(color = 'rgba(160, 112, 20, 0.8)'),)

data = [trace1, trace2]
layout = go.Layout(
    xaxis2=
        dict(domain=[0.6, 0.95],
        anchor='y2',),     # 2 ıncı plot cizdirir
    yaxis2=
        dict(domain=[0.6, 0.95],
        anchor='x2',),     # 2 ıncı plot cizdirir
    title = 'Income and Teaching vs World Rank of Universities')

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
