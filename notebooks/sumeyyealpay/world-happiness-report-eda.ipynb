```python
data_20=pd.read_csv('../input/world-happiness-report/2020.csv')
data_19=pd.read_csv('../input/world-happiness-report/2019.csv')
data_18=pd.read_csv('../input/world-happiness-report/2018.csv')
data_17=pd.read_csv('../input/world-happiness-report/2017.csv')
data_16=pd.read_csv('../input/world-happiness-report/2016.csv')
data_15=pd.read_csv('../input/world-happiness-report/2015.csv')
```

```python
data_15['Year']=2015
data_16['Year']=2016
data_17['Year']=2017
data_18['Year']=2018
data_19['Year']=2019
data_20['Year']=2020
```

```python
data_15=data_15.rename(columns={'Family':'Social Support'})
data_16=data_16.rename(columns={'Family':'Social Support'})
data_17=data_17.rename(columns={'Happiness.Rank':'Happiness Rank', 'Happiness.Score':'Happiness Score',
                              'Economy..GDP.per.Capita.':'Economy (GDP per Capita)','Family':'Social Support','Health..Life.Expectancy.':'Health (Life Expectancy)',
                                'Trust..Government.Corruption.':'Trust (Government Corruption)'})
data_18=data_18.rename(columns={'Overall rank':'Happiness Rank','Country or region':'Country','Score':'Happiness Score',
                               'GDP per capita':'Economy (GDP per Capita)','Family':'Social Support','Healthy life expectancy':'Health (Life Expectancy)',
                                'Freedom to make life choices':'Freedom','Perceptions of corruption':'Trust (Government Corruption)'})
data_19=data_19.rename(columns={'Overall rank':'Happiness Rank','Country or region':'Country','Score':'Happiness Score',
                               'GDP per capita':'Economy (GDP per Capita)','Social support':'Social Support','Healthy life expectancy':'Health (Life Expectancy)',
                                'Freedom to make life choices':'Freedom','Perceptions of corruption':'Trust (Government Corruption)'})
data_20['Happiness Rank']=range(1, len(data_20.index)+1)
data_20=data_20.rename(columns={'Country name':'Country','Ladder score':'Happiness Score','Logged GDP per capita':'Economy (GDP per Capita)','Social support':'Social Support','Healthy life expectancy':'HealthHealth (Life Expectancy)',
                               'Freedom to make life choices':'Freedom','Perceptions of corruption':'Trust (Government Corruption)'})
```

```python
data=pd.concat([data_15,data_16,data_17,data_18,data_19,data_20])
data=data[['Happiness Score','Country','Year']]
data.head()
```

```python
x_15=data_15['Country']
y_15=data_15['Happiness Score']
x_16=data_16['Country']
y_16=data_16['Happiness Score']
x_17=data_17['Country']
y_17=data_17['Happiness Score']
x_18=data_18['Country']
y_18=data_18['Happiness Score']
x_19=data_19['Country']
y_19=data_19['Happiness Score']
x_20=data_20['Country']
y_20=data_20['Happiness Score']
plt.figure(figsize=(50,40))
plt.plot(y_15,x_15,label='2015')
plt.plot(y_16,x_16,label='2016')
plt.plot(y_17,x_17,label='2017')
plt.plot(y_18,x_18,label='2018')
plt.plot(y_19,x_19,label='2019')
plt.plot(y_20,x_20,label='2020')
plt.legend()
xlabel='Country'
ylabel='Happiness Score'
plt.title('Yıllara göre ülkelerin mutluluk skorları',fontsize=30)
plt.show()
```

```python
mean=data.groupby(['Country']).mean().sort_values(by='Happiness Score',ascending=False)
mean
```

```python
median=data.groupby(['Country']).median().sort_values(by='Happiness Score',ascending=False)
median
```

```python
s=list(set(mean['Happiness Score']).intersection(set(median['Happiness Score'])))
pd.isnull(s)
```
