# Load IRIS dataset and shuffle the dataframe:

```python
from math import sqrt
import pandas as pd

df = pd.read_csv("/kaggle/input/iris-flower-dataset/IRIS.csv") 
#shuffle the dataframe
df = df.sample(frac = 1)   
```

```python
df.head()
```

Convert dataframe to numpy array**

```python
dataset = df.to_numpy()
```

# point_distance(): function for calculating eucledian distance between two data points
# K_neighbors(): function for getting k nearest neighbors
# Predict(): function for making prediction with KNN algorithm


```python
def point_distance(row1, row2):
	distance = 0.0
	for i in range(len(row1)-1):
		distance += (row1[i] - row2[i])**2
	return sqrt(distance)

     
def K_neighbors(train, test_row, num_neighbors):
	distances = list()
	for train_row in train:
		dist = point_distance(test_row, train_row)
		distances.append((train_row, dist))
	distances.sort(key=lambda tup: tup[1])
	neighbors = list()
	for i in range(num_neighbors):
		neighbors.append(distances[i][0])
	return neighbors


def predict(train, test_row, num_neighbors):
	neighbors = K_neighbors(train, test_row, num_neighbors)
	output_values = [row[-1] for row in neighbors]
	prediction = max(set(output_values), key=output_values.count)
	return prediction    
```

# Input appropriate value of K

```python
k_neighbors = int(input("K=?"))
```

# Show K nearest neighbors

```python
print("K=",k_neighbors,"nearest neighbors:")
neighbors = K_neighbors(dataset, dataset[0], k_neighbors)
for neighbor in neighbors:
	print(neighbor)
```
