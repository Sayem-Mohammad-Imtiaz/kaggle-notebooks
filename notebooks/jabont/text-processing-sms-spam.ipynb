```python
import pandas as pd
import sklearn.metrics as metrs
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
data = pd.read_csv('//kaggle//input//sms-spam-collection-dataset//spam.csv',header=0,encoding='latin-1',usecols=[0,1])

print(data.head(10))

#------------------select y data
y=data['v1'].replace(['ham','spam'],[0,1])
print(y.head(10))
print(y.shape)

############## select X features data
wordvec=CountVectorizer().fit_transform(data['v2'])
X=wordvec.toarray()
print(X[:10])
print(X.shape)
```

```python
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,random_state=42)
print(Xtrain.shape)
print(Xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
#------------------Naive Model-----------
nb = MultinomialNB().fit(Xtrain,ytrain)
print('TestScore-------------',nb.score(Xtest,ytest))
print('TrainScore-------------',nb.score(Xtrain,ytrain))
```

```python
#-----------------Model-evaluation------------
metrs.plot_confusion_matrix(nb,Xtest,ytest,display_labels=['ham','spam'])
metrs.plot_roc_curve(nb,Xtest,ytest)
```
