```python
cols =['Sex','BP','Cholesterol']

df_new=pd.get_dummies(df,columns=cols,drop_first=True)

df_new
```

```python
y=df['Drug']

X=df_new.drop(['Drug'],axis=1)

X

```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)


import lightgbm as lgb
clf = lgb.LGBMClassifier()
clf.fit(X_train, y_train)
preds = clf.predict(X_test)
acc_lgbm=accuracy_score(y_test, preds)
print("XGBoost's prediction accuracy is: %3.2f" % (acc_lgbm))
```

```python
from  xgboost import XGBClassifier
import time


xgb = XGBClassifier(n_estimators=100)
training_start = time.perf_counter()
xgb.fit(X_train, y_train)
training_end = time.perf_counter()
prediction_start = time.perf_counter()
preds = xgb.predict(X_test)
prediction_end = time.perf_counter()
acc_xgb=accuracy_score(y_test, preds)
print("XGBoost's prediction accuracy is: %3.2f" % (acc_xgb))

```
