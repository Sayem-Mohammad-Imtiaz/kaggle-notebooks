```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.model_selection import cross_val_score
```

```python
data=pd.read_csv("../input/iris-flower-dataset/IRIS.csv")
```

```python
data.columns
```

```python
data.head()
```

```python
data.species.value_counts()
```

```python
X=data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
```

```python
y=data[['species']]
```

```python
X
```

```python
y
```

```python
model=svm.SVC()
accuracy=cross_val_score(model,X,y.values.ravel(),scoring='accuracy',cv=10)
print(accuracy)

#y.values.ravel() was used to remove the conda warning "A column-vector y was passed when a 1d array was expected"
#.values will give the values in an array. (shape: (n,1)

#.ravel will convert that array shape to (n, )

#getting mean of each fold
print("Accuracy of Model with Cross Validation is:", accuracy.mean()*100)
```
