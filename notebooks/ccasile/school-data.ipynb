Import Pandas:

```python
import pandas as pd

data = pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
data.head()
```

Split X and y, then convert y's to 1's and 0's

```python
y = data['test preparation course']
X = data
X.drop(columns=['test preparation course'], inplace=True)
y = y.map(dict(completed=1, none=0))
print(y)
```

Check for any null values:

```python
data.isna().sum()
```

Get dummy variables for string columns:

```python
X = pd.get_dummies(X, drop_first=True)
X.head()
```

Check for correlation between X values:

```python
X.corr()
```

Math, reading and writing all seemed to be highly correlated so 2 of the 3 are dropped:

```python
X.drop(columns=['math score', 'reading score'], inplace=True)
```

All values are normalized to between 0 and 1:

```python
from sklearn import preprocessing

cols = X.columns
x = X.values 
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
X = pd.DataFrame(x_scaled, index=X.index, columns=cols)
X.head()
```

Train and test values are created:

```python
from sklearn.model_selection import train_test_split
X = X.astype(float)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

Model is fitted and predictions are made:

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state=0)

logreg.fit(X_train,y_train)

predictions = logreg.predict(X_test)
```

```python

from sklearn.metrics import accuracy_score

accuracy_score(y_test, predictions)
```

Confusion Matrix:

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, predictions)
```

The below code outputs a linear regression output:

```python
import statsmodels.api as sm
from scipy import stats

X2 = sm.add_constant(X)
est = sm.OLS(y, X2)
est2 = est.fit()
print(est2.summary())
```
