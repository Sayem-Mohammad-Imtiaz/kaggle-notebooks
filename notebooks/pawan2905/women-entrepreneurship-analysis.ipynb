```python
crosstab = pd.crosstab(df['European Union Membership'], df['Level of development'])

crosstab.style.background_gradient(cmap='YlGnBu', low=0, high=0, axis=0, subset=None)
```

```python
from scipy import stats
```

- H0: There's no significant relationship between a countrys' development and EU Membership.
- Ha: There's a significant relationship between a countrys' development and EU Membership.


```python
stats.chi2_contingency(crosstab)
```

```python
data_eu=df[df['European Union Membership']=='Member']
data_not_eu=df[df['European Union Membership']=='Not Member']
```

```python
def mann_whitney_u_test(distribution_1, distribution_2):
    """
    Perform the Mann-Whitney U Test, comparing two different distributions.
    Args:
       distribution_1: List. 
       distribution_2: List.
    Outputs:
        u_statistic: Float. U statisitic for the test.
        p_value: Float.
    """
    u_statistic, p_value = stats.mannwhitneyu(distribution_1, distribution_2)
    return u_statistic, p_value


#### MAIN FUNCTION ####
#Perform the Mann-Whitney U Test on the two distributions
mann_whitney_u_test(list(data_eu['Women Entrepreneurship Index']), list(data_not_eu['Women Entrepreneurship Index']))
```

```python
mann_whitney_u_test(list(data_eu['Entrepreneurship Index']), list(data_not_eu['Entrepreneurship Index']))
```

Is there a statistically significant relationship between Women's Entrepreneurship Index and Global Entrepreneurship Index values?

```python
# Chi2 
(chi2, p, dof,_) = stats.chi2_contingency([crosstab.iloc[0].values, crosstab.iloc[1].values])
values = [chi2, p, dof]

pd.DataFrame(values, index = ['chi2', 'p', 'dof'], columns = {'values'})
```

> There is a statistically significant relationship between Women's Entrepreneurship Index and Global Entrepreneurship Index values.

> Cramer V indicates the strength of the association. The closer to 1 the stronger the association is.

```python
cramer_v_test = np.sqrt(chi2/(df.shape[0]))
cramer_v_test
```

> Spearman Correlation :
Spearman correlation helps us to find the degree of association of two-variables. It evaluates a monotonic relationship between a continuous variable or ordinal variable

```python
df[['Women Entrepreneurship Index','Entrepreneurship Index']].corr(method='spearman').iloc[0,1]
```
