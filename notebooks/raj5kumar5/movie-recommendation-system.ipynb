```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
netflix = pd.read_csv('../input/netflix-shows/netflix_titles.csv',index_col='title')
```

```python
netflix.head()
```

```python
netflix.columns
```

```python
required_data=netflix[['type','country','rating','listed_in']]
```

```python
required_data.head()
```

```python
required_data.info()
```

```python
required_data.dropna(inplace=True)
```

```python
required_data.info()
```

# First Method # 

```python
required_data1=required_data.copy()
```

```python
for i in list(required_data1.columns):
    LE=LabelEncoder()
    LE.fit(required_data1[i])
    required_data1[i]=LE.transform(required_data1[i])
```

```python
required_data1.head()
```

```python
required_data1=required_data1.T
```

```python
required_data1.head()
```

```python
#Enter movie name

search=required_data1['3%']
```

```python
related=required_data1.corrwith(search)
```

```python
related.sort_values(ascending=False).head(10)
```

# Second Method # 

```python
required_data2=pd.DataFrame()
for i in list(required_data.columns):
    required_data2[pd.get_dummies(required_data[i]).columns]=(pd.get_dummies(required_data[i]))
```

```python
required_data2.head()
```

```python
required_data2=required_data2.T
```
