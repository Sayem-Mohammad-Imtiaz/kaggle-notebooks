```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/kaggle/input/co2-ghg-emissionsdata/co2_emission.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#since code has a no. of null values & we also don't require it in our analysis, we will drop it.
df.drop(["Code"],axis=1,inplace=True)
```

```python
df.head()
```

```python
#renaming columns
df=df.rename(columns={"Annual COâ‚‚ emissions (tonnes )":"emission_index"})
```

```python
df1=df.copy()
```

1. COUNTRY WITH HIGHEST CO2 EMISSIONS ACROSS ALL THE YEARS.

```python
df1.drop(["Year"],axis=1,inplace=True)
```

```python
df1.head()
```

```python
df1=df1.groupby("Entity").sum()
```

```python
df1.head()
```

```python
df1.reset_index(inplace=True)
```

```python
df1.head()
```

```python
df1=df1.sort_values("emission_index",ascending=False)
```

```python
dfx=df1.head(20)
```

```python
plt.figure(figsize=(20,6))
sns.set_style('whitegrid')
sns.barplot(x=dfx["Entity"],y=dfx["emission_index"])
plt.xticks(rotation=90)
```

KEY POINT:

UNITED STATES STANDS AT FIRST IN EMISSION INDEX RANKINGS. IN FACT IT HAS MORE EMISSION INDEX IN COMPARISION
TO ANY OTHER CONTINENTS !!



2. HIGHEST CO2 EMISSION IN 21ST CENTURY

```python
df.head()
```

```python
df["Year"]=df["Year"].astype(int)
```

```python
df_20=df[df["Year"] > 1999]
```

```python
df_20.drop(["Year"],axis=1,inplace=True)
```

```python
df_20=df_20.groupby("Entity").sum()
```

```python
df_20.reset_index(inplace=True)
```

```python
df_20=df_20.sort_values("emission_index",ascending=False)
```

```python
dfx=df_20.head(20)
```

```python
plt.figure(figsize=(20,6))
sns.set_style('whitegrid')
sns.set_palette("summer")
sns.barplot(x=dfx["Entity"],y=dfx["emission_index"])
plt.xticks(rotation=90)
```

KEY POINT :

CHINA BEATS EVERYONE IN 21ST CENTURY IN EMISSION INDEX

3. YEAR WISE CO2 EMISION OF ALL COUNTRY

```python
df.head()
```

```python
#we will compare co2 emission of 3 countries
```

```python
df_ind=df[df["Entity"]=="India"]
df_ind.reset_index(drop=True,inplace=True)
```

```python
df_usa=df[df["Entity"]=="United States"]
df_usa.reset_index(drop=True,inplace=True)
```

```python
df_chn=df[df["Entity"]=="China"]
df_chn.reset_index(drop=True,inplace=True)
```

```python
df_jpn=df[df["Entity"]=="Japan"]
df_jpn.reset_index(drop=True,inplace=True)
```

```python
df_ger=df[df["Entity"]=="Germany"]
df_ger.reset_index(drop=True,inplace=True)
```

```python
plt.figure(figsize=(20,6))
plt.plot(df_ind["emission_index"],color="orange",label="India")
plt.plot(df_usa["emission_index"],color="blue",label="USA")
plt.plot(df_chn["emission_index"],color="green",label="China")
plt.plot(df_jpn["emission_index"],color="yellow",label="Japan")
plt.plot(df_ger["emission_index"],color="red",label="Ger")
plt.legend()
```
