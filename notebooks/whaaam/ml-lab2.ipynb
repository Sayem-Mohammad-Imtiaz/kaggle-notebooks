## indexing


```python
df.iloc[2]
```

here we have done indexing using iloc which is giving us information about 2nd row . Code for iloc is dataframe.iloc[row,column] and since we mentioned 2nd row we got info about second row in all columns


```python
df.loc[:,'Lat']
```

Here we have used loc in indexing which is similar to iloc but instead of indexes it takes names, now here we have asked it to show all the rows of the column 'lat'(we used : symbol to get all the rows )

## Indexing multiple rows/columns

```python
df.iloc[[2,3,4]]
```

Here we got access to multiple rows using iloc command and it is similar to what we used before , the difference here is we mentioned all the rows we wanted . Let's see the same command with loc in columns

```python
df.loc[:,['Lat','Long']]
```

## Selecting subset of data 

```python
(df.Lat =='42.5063').head()
```

Here we got a result as bool which means if it is true we have that latitude

```python
df[df.Date == "2020-01-22"]
```

Here we get information of every country at 22 Jan in form of dataframe 

You can also have multiple conditions here 

```python
df[(df.Date == "2020-01-22") | (df.Long == '67.709953')].head()
```

Here we have mentioned an or condition i.e whether date should be this or longitude should be the one mentioned 

## Visualizing the data

### Univiriate plots 

```python
df['Country/Region'].value_counts().head(10).plot.bar()
```

```python
(df['Country/Region'].value_counts().head(10) / len(df)).plot.bar()
```

### Line charts 

```python
df['Country/Region'].value_counts().sort_index().plot.line()
```

### area chart

```python
df['Country/Region'].value_counts().sort_index().plot.area()
```

### Histogram 

```python
df['Lat'].plot.hist()
```

## Bivariate plots 

```python
df[df['Confirmed'] < 100].plot.scatter(x='Confirmed', y='Lat')
```

## Hexplots

```python
df[df['Confirmed'] < 10000].plot.hexbin(x='Confirmed', y='Long')
```
