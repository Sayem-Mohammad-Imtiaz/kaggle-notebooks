```python
# importing libraires
%matplotlib inline                 

import pandas as pd                # Implemennts milti-dimensional array and matrices
import numpy as np                 # For data manipulation and analysis
import matplotlib.pyplot as plt    # Plotting library for Python programming language and it's numerical mathematics extension NumPy
import seaborn as sns              # Provides a high level interface for drawing attractive and informative statistical graphics
import bar_chart_race as bcr       # For running bar chart, over time period
```

```python
covid = pd.read_csv("../input/covid19-in-india/covid_19_india.csv")
covid.head()
```

```python
covid.info()
```

```python
covid.describe()
```

```python
covid.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'], axis=1, inplace=True)
covid.drop(['Cured','Deaths'], axis=1, inplace=True)
covid.rename(columns = {'State/UnionTerritory':'States'}, inplace=True)
covid.columns
```

```python
covid['Date'] = pd.to_datetime(covid['Date'])
```

```python
import datetime
covid.drop(covid[covid['Date']< datetime.datetime(2020, 3, 1) ].index, inplace=True)
```

```python
covid
```

```python
covid['States'].unique()
```

```python
covid['States'].replace({"Maharshtra***" : "Maharashtra", "Chandigarh***" : "Chandigarh",
                         "Punjab***" : "Punjab", "Telengana" : "Telangana", "Telengana***" : "Telangana", 
                         "Telangana***" : "Telangana",
                         "Daman & Diu" : "Dadra and Nagar Haveli and Daman and Diu",
                         "Dadar Nagar Haveli" : "Dadra and Nagar Haveli and Daman and Diu",
                        "Maharashtra***" : "Maharashtra"}, inplace = True)
covid = covid[(covid['States'] != 'Cases being reassigned to states') 
              & (covid['States'] != 'Unassigned')]
```

```python
covid["States"].unique()
```

```python
covid['States'].nunique()
```

```python
india = pd.pivot_table(covid, values = 'Confirmed', index = 'Date', columns = 'States')
india.fillna(0)
```

```python
bcr.bar_chart_race(df=india,title="Covid 19 confirmed cases",n_bars=10,filter_column_colors=True,figsize=(5,3))
```
