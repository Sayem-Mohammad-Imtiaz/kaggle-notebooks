```python
df_2016 = pd.read_csv('/kaggle/input/world-happiness/2016.csv')
df_2016.head(10)
```

```python
df_2016_proc = df_2016[['Country','Region', 'Happiness Score', 'Economy (GDP per Capita)', 'Health (Life Expectancy)', 'Trust (Government Corruption)','Generosity','Family','Freedom']]
df_2016_sample = df_2016_proc.loc[df_2016_proc['Region'].isin(['Western Europe', 'North America','Southeastern Asia','Central and Eastern Europe', 'Eastern Asia'])]
df_2016_sample.head(10)
```

```python
def bkapp(doc):
    #colors = ["red", "blue", "green", "orange", "purple"]
    index_cmap = factor_cmap('Region', palette='Spectral5', factors=sorted(df_2016_sample.Region.unique()))

    hover = HoverTool(tooltips=[
        ("Country ", "@Country"),
        ("Economy (GDP per Capita)", "@{Economy (GDP per Capita)}"),
        ('Happiness Score', '@{Happiness Score}'),
    ])

    TOOLS="tap,pan,wheel_zoom,zoom_in,zoom_out"
    p = figure(tools=[hover, TOOLS], title="Scatter Plot of GDP & Happiness_Score Index in 2016")
    p.scatter('Economy (GDP per Capita)', 'Happiness Score',source=df_2016_sample, fill_color=index_cmap, size=15, legend_field='Region')
    p.xaxis.axis_label = 'GDP'
    p.yaxis.axis_label = 'Happiness Score'
    p.legend.location = "top_left"
    p.width=1000
    p.height=700

    def my_dropdown_input_handler(attr, old, new):
        p.renderers = []
        p.scatter(str(new), 'Happiness Score',source=df_2016_sample, fill_color=index_cmap, size=15, legend_field='Region')
        p.xaxis.axis_label = str(new)


    menu = Select(title="Option:", value="foo", options=["Economy (GDP per Capita)", "Family", "Health (Life Expectancy)", "Trust (Government Corruption)", "Generosity"])
    menu.on_change("value", my_dropdown_input_handler)

    doc.add_root(column(menu, p))
```

```python
show(bkapp)
```
