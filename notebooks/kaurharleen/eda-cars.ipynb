```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes = True)
```

```python
df = pd.read_csv('../input/cardataset/data.csv')
df.head()
```

```python
df.tail()
```

```python
df.dtypes
```

```python
df = df.drop(['Engine Fuel Type', 'Number of Doors', 'Market Category', 'Vehicle Size', 'Vehicle Style'], axis =1)
df.head(5)
```

```python
df = df.rename(columns= {
    'Engine HP' : 'HP',
    'Engine Cylinders' : 'Cylinders',
    'Transmission Type' : 'Transmission',
    'Driven_Wheels' : 'Driving Mode',
    'highway MPG' : 'MPG-H',
    'city mpg' : 'MPG-C',
    
})
df.head(5)
```

```python
df.shape
```

```python
duplicate_rows_df = df[df.duplicated()]
print('number of duplicated rows:' ,duplicate_rows_df.shape)
```

```python
df.count()
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
df = df.dropna()
df.count()
```

```python
print(df.isnull().sum())
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3-Q1
print(IQR)
```

```python
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```

```python
sns.boxplot(x=df['MSRP'])
```

```python
df.Make.value_counts().nlargest(40).plot(kind = 'bar',figsize = (20,10))
plt.title('No. of cars by different brands')
plt.xlabel('Names of Brands')
plt.ylabel('Number of Cars')
```

```python
plt.figure(figsize = (10,6))
c = df.corr()
sns.heatmap(c,cmap= 'YlGnBu', annot = True)
c
```

```python
fig,ax = plt.subplots(figsize = (10,6))
ax.scatter(df['HP'], df['MSRP'], c = 'Red')
ax.set_xlabel('Horsepower')
ax.set_ylabel('Price')
plt.show()
```
