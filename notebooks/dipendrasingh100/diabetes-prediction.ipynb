```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split,cross_val_score,KFold
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,precision_score,recall_score
```

```python
#importing dataset
df = pd.read_csv("../input/pima-indians-diabetes-database/diabetes.csv")
```

```python
# removeing BloodPressure,SkinThickness columns
df.drop(["BloodPressure","SkinThickness"],axis=1,inplace=True)
```

```python
#indenpendent and dependent variables
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X.shape
```

```python
y.shape
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
X_scale = scaler.transform(X)
```

```python
clf = SVC(kernel='linear')
```

```python
kfold = KFold(n_splits=3,shuffle=True,random_state=2)
score = cross_val_score(clf,X_scale,y,cv=kfold)
```

```python
score
```

```python
print("Overall Accuracy of Model - ",score.mean())
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_scale,y,stratify=y,random_state=2)
```

```python
clf.fit(X_train,y_train)
```

```python
pred = clf.predict(X_test)
```

```python
#Confussion matrix
confusion_matrix(pred,y_test)
```

```python
precision = precision_score(pred,y_test)
recall = recall_score(pred,y_test)
```

```python
print("Accuracy on train set - ",clf.score(X_train,y_train))
print("Accuracy on test set - ",clf.score(X_test,y_test))
```
