## split data

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn import metrics

x_df = train_df.iloc[:,:-1]
y_df = train_df['target']
X_train, X_test, y_train, y_test = train_test_split(x_df, y_df,train_size = 0.9,random_state = 0)
```

## deal with the unalanced dataset

```python
from imblearn.over_sampling import SMOTE 
```

```python
#oversampling using SMOT
# transform the dataset
oversample = SMOTE()
X_train, y_train = oversample.fit_resample(X_train, y_train)
```

## implement the model

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import plot_roc_curve
from sklearn.metrics import confusion_matrix
```

### SVM model without standarization

```python
from sklearn.svm import SVC

clf = SVC(kernel = 'rbf', C=10).fit(X_train, y_train)

print("accurary on training set: ",clf.score(X_train, y_train))
print("accurary on test set: ",clf.score(X_test, y_test))

plot_roc_curve(clf, X_test, y_test, name = 'SVM')
```

### SVM model with standarization

```python
from sklearn.preprocessing import StandardScaler

scaler_std = StandardScaler()
X_train_std = scaler_std.fit_transform(X_train)
X_test_std = scaler_std.transform(X_test)
```

```python
from sklearn.svm import SVC
clf_std = SVC().fit(X_train_std, y_train)
print("accurary on training set: ",clf_std.score(X_train_std, y_train))
print("accurary on test set: ",clf_std.score(X_test_std, y_test))

plot_roc_curve(clf_std, X_test_std, y_test, name = 'SVM_std')
```

### DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier().fit(X_train, y_train)

print("accurary on training set: ",dtc.score(X_train, y_train))
print("accurary on test set: ",dtc.score(X_test, y_test))

plot_roc_curve(dtc, X_test, y_test, name = 'DT')
```

```python
from sklearn.model_selection import RandomizedSearchCV

#https://www.analyticsvidhya.com/blog/2020/03/beginners-guide-random-forest-hyperparameter-tuning/


# Number of trees in random forest
n_estimators = [800, 1000, 1200, 1400, 1600, 1800, 2000]

# Number of features to consider at every split
max_features = [4, 5, 6]

#The sub-sample size used to build each tree
max_samples = [0.2, 0.5, 0.7]

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(4, 20, num = 9)]
max_depth.append(None)

# Minimum number of samples required to split a node
min_samples_split = [2, 4]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2]


random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_samples' : max_samples,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
print(random_grid)

```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf_random.fit(X_train, y_train)
```

```python
rf_random.best_params_
```

```python
print("accurary on training set: ",rf_random.score(X_train, y_train))
print("accurary on test set: ",rf_random.score(X_test, y_test))


plot_roc_curve(rf_random, X_test, y_test, name = 'rf_random')
```
