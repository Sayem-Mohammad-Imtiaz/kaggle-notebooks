```python
import pandas as pd 
train = pd.read_csv("../input/mushroom-classification/mushrooms.csv")
print(train.head())
```

```python
from sklearn import preprocessing
labelencoder = preprocessing.LabelEncoder()
labels = list(train.columns)
for label in labels:
    train[label] = labelencoder.fit_transform(train[label])
train_y = train[['class']]
train.drop(['class'], axis = 1, inplace = True)
print(train.head())
print(train_y.head())
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(train, train_y, random_state = 0)
print(x_train.shape)
print(x_val.shape)
print(y_train.shape)
print(y_val.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_val, model.predict(x_val))
print(mae)
```
