Which features are categorical?

```python
for c in features:
    print(f"feature {c}")
    print(df[c].nunique())
```

```python
categorical_features = [c for c in features if df[c].nunique() < 204]
print(categorical_features)
```

Which features are numerical?

```python
numerical_feats = [c for c in features if c not in categorical_features]
```

```python
print(numerical_feats)
```

How many unique title of emergency codes are there?

```python
# df["title"].unique()
```

```python
def extract_title(text):
    code = text.split(':')[0]
    return code
```

```python
df["title_code"] = df["title"].map(extract_title)
```

```python
df["title_code"].nunique()
```

What is the most common Reason for a 911 call based off of this new column?

```python
df["title_code"].value_counts()
```
