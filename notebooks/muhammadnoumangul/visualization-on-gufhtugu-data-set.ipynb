```python
from itertools import chain
lens = df['book_name'].str.split('/').map(len)
df = pd.DataFrame({'order_no': np.repeat(df['order_no'], lens),'order_status': np.repeat(df['order_status'], lens),
                    'book_name': chainer(df['book_name']),'order_date': np.repeat(df['order_date'], lens),'city': np.repeat(df['city'], lens)})
df.reset_index(inplace=True,drop=True)
df
```

```python
sns.countplot(x='book_name',data=df)
```
