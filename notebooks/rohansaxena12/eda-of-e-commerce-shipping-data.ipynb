```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x=df.Warehouse_block,y=df.Cost_of_the_Product).set_title("Cost_of_the_Product VS Warehouse_block BarPlot")
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.barplot(x=df.Warehouse_block,y=df.Cost_of_the_Product,hue=df.Product_importance).set_title("Cost_of_the_Product VS Warehouse_block BarPlot On the basis of Product Importance")
```

```python
sns.boxplot(x=df.Mode_of_Shipment,y=df.Weight_in_gms).set_title('Weight_in_gms VS Mode_of_shipment box plot')
```

```python
sns.lineplot(x='Warehouse_block',y='Customer_rating',hue='Reached.on.Time_Y.N',data=df).set_title('Warehouse_block VS Customer_rating lineplot')
```
