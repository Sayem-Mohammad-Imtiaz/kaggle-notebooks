```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv(r"../input/indian-food-101/indian_food.csv")
df.head()
```

```python
df.columns
```

```python
df["name"].unique()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.head()
```

```python
np.shape(df)
```

```python
df["prep_time"]
```

```python
df["cook_time"]
```

```python
df['prep_time'] = df['prep_time'].replace(-1, 0)
df['cook_time'] = df['cook_time'].replace(-1, 0)
```

```python
df
```

```python
df["prep_time"].mean()
```

```python
df["cook_time"].mean()
```

```python
df['prep_time'] = df['prep_time'].replace(0, 31)
df['cook_time'] = df['cook_time'].replace(0, 34)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["region"].tail(10)
```

```python
df = df[df.region != '-1']
df = df[df.state != '-1']
df = df[df.flavor_profile != '-1']
```

```python
df["region"].tail(10)
```

```python
b=sns.catplot(data=df, kind="bar", x="diet", y="prep_time", hue="region")
b.fig.suptitle('Prep time Vs diet plot on the basis of Region')
```

**Graph information** 1.North indian Non vegetarian food take more preperation time 2.In vegetarian North east
food Take more preperation time 3.In non vegetarian East food take less Prep time 4.In Vegetarian Central
food take less prep time

```python
df.head()
```

```python
b=sns.catplot(data=df, kind="bar", x="diet", y="cook_time", hue="region")
b.fig.suptitle('Cook time Vs diet plot on the basis of Region')
```

**Graph Information: 1.Non Vegetarian south indian food take more cook time 2.Non vegetaian food Take More
cook_time**

BY LOOKING ON ABOVE TWO GRAPHS IT IS CLEARLY VISIBLE THAT NON VEGETARIAN FOOD TAKE
