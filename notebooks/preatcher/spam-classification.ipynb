```python
from nltk.stem  import SnowballStemmer
st = SnowballStemmer(language = 'english')
def stemmer_snowball(sentence):
    return (' '.join([st.stem(word) for word in sentence.split(' ')]))
text.apply(stemmer_snowball)
    
```

### Creating Text Preprocessing Pipeline

```python
from wordcloud import WordCloud, STOPWORDS
from nltk.corpus import stopwords
print(len(STOPWORDS))
print(len(stopwords.words('english')))
```

```python
def process_text(text):
    import re
    import nltk
    from textblob import TextBlob
    from nltk.corpus import stopwords
    from nltk.stem import SnowballStemmer
    from textblob import Word
    from nltk.util import ngrams
    from wordcloud import WordCloud, STOPWORDS
    from nltk.tokenize import word_tokenize
    stemmer = SnowballStemmer(language = 'english')
    row = text
    # lowercasing
    row = row.lower()
    # Remove non alphanumeric symbols white spaces
    row = re.sub(r'[^\w\s%]+','',row)
    row = re.sub(r'[\d]+',' digit ',row)
    row = re.sub(r'[%]+',' p ',row)
    row =  [stemmer.stem(word) for word in row.split() if word not in STOPWORDS]   
    return row
    
```

```python
sample_text = pd.Series(['This is abhishek_jaiswal9957   87897per','787985%^ and we know usf',
                        'this &^&*^* '])

from sklearn.feature_extraction.text import TfidfVectorizer
cnt = CountVectorizer( analyzer=process_text)
cnt.fit(sample_text)
pd.DataFrame(data = cnt.transform(sample_text).toarray(), columns = cnt.get_feature_names())
```

### Generating N-grams
in count vectorizer method we have a draw back that whenever we create features by using count vectorizer method we take single words at a time 
now if "not good " means bad but in count vectorizer it will get treated as good since good will have a feature 

to rule out this problem we have n_grams here we look for 2/3/4 words at a time N words at a time


```python
from sklearn.feature_extraction.text import CountVectorizer
cnt_ngram = CountVectorizer(ngram_range=(1,2), preprocessor = process_text)
# tokenizing
cnt_ngram.fit(text)

# encoding features 
features = cnt_ngram.transform(text).toarray()
# features names
features_name = cnt_ngram.get_feature_names()
```

```python
pd.DataFrame(data = features, columns = features_name)
```

### TF-IDF count vectorizer
* #### The whole idea of tf_idf is to reflect on how important a word is to a document in a collection and hence normalizing words appeared frequently in all the documents 
* #### Let's say particular word is appearing in all the documents of the corpus, then it will achieve higher importance in our previous method (count vectorizer) thats bad

* #### Term Frequency(TF) is simply the ratio of the count of a word present in a sentence, to the length of the sentence


```python
text = ['The quick brown fox jumped over the lazy dog.', 'The Dog','The Fox']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# create the vectorizer
cnt_idf = TfidfVectorizer(stop_words='english', preprocessor = process_text, ngram_range=(1,2))


# tokenize and build vocab
cnt_idf.fit(text)

print(cnt_idf.vocabulary_)
print(cnt_idf.idf_)
```

## Text Processing function

```python
def process_token(text):
    import re
    import nltk
    from textblob import TextBlob
    from nltk.stem import SnowballStemmer
    from textblob import Word
    from wordcloud import WordCloud, STOPWORDS
    from nltk.tokenize import word_tokenize
    stemmer = SnowballStemmer(language = 'english')
    row = text
    # lowercasing
    row = row.lower()
    # Remove non alphanumeric symbols white spaces
    row = re.sub(r'[^\w\s%]+','',row)
    row = re.sub(r'[\d]+',' 9 ',row)
    row = re.sub(r'[%]+',' p ',row)
    row =  [stemmer.stem(word) for word in row.split() if word not in STOPWORDS]
    return row

def process_text(text):
    import re
    import nltk
    from textblob import TextBlob
    from nltk.stem import SnowballStemmer
    from textblob import Word
    from wordcloud import WordCloud, STOPWORDS
    from nltk.tokenize import word_tokenize
    stemmer = SnowballStemmer(language = 'english')
    row = text
    # lowercasing
    row = row.lower()
    # Remove non alphanumeric symbols white spaces
    row = re.sub(r'[^\w\s%]+','',row)
    row = re.sub(r'[\d]+',' 9 ',row)
    row = re.sub(r'[%]+',' p ',row)
    row =  ' '.join([stemmer.stem(word) for word in row.split() if word not in STOPWORDS])
    return row
```

```python
# splitting the data to check our final model 
from sklearn import model_selection
X_train,X_test,y_train,y_test = model_selection.train_test_split(sms['text'],sms.label_num,
stratify = sms.label_num, test_size = 0.2, random_state = 42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfTransformer
from sklearn.naive_bayes import GaussianNB,MultinomialNB

from sklearn.pipeline import Pipeline
vector = CountVectorizer(analyzer=process_token)
classifier  = Pipeline([('feature_generation', vector), ('model',MultinomialNB(alpha = 1.9))])
```

```python
classifier.fit(X_train,y_train)
```
