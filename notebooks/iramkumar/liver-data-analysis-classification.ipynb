# Replacing Categorical Variable Gender

```python
male = pd.get_dummies(df_liver['Gender'], drop_first = True)
    
df_liver.drop(['Gender'], axis = 1,inplace = True)
df_liver = pd.concat([df_liver, male], axis = 1)
```

```python
plt.figure(figsize=(20,10)) 
sns.heatmap(df_liver.corr(), annot=True) 
plt.show()
```

# Setting up Model Parameters

```python
X = df_liver.drop('Dataset', axis = 1).values
y = df_liver['Dataset'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,random_state = 0)
X_train.shape
```

# Feature Scaling the values

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
```

# Running Various Classification Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC 
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
models = [['Logistic Regresssion :',LogisticRegression()],
           ['Decision Tree Classification :', DecisionTreeClassifier()],
           ['RandomForest Classifier :',RandomForestClassifier(n_estimators = 200,criterion = 'entropy')],
           ['KNeighbours Classifier:', KNeighborsClassifier(n_neighbors = 5)],
           ['SVM :', SVC()],
           ['NB Classifier :', GaussianNB()]]
```

```python
for name_of_model,model_parms in models:
    curr_model = model_parms
    curr_model.fit(X_train, y_train)
    curr_predict = curr_model.predict(X_test)
    print(50*"=")
    print(10*" ",name_of_model,20*" ")
    print(50*"=")
    print("Accuracy Score  :", accuracy_score(y_test, curr_predict))
    print("Confusion Matrix :\n",confusion_matrix(y_test, curr_predict))
    print("Classification Report :\n", classification_report(y_test,curr_predict))
```
