```python
male = pd.get_dummies(df_sp['gender'], drop_first = True)
df_sp.drop(['gender'], axis=1, inplace=True)
df_sp = pd.concat([df_sp, male], axis=1)
```

```python
ever_married = pd.get_dummies(df_sp['ever_married'], drop_first = True)
df_sp.drop(['ever_married'], axis=1, inplace=True)
df_sp = pd.concat([df_sp, ever_married], axis=1)
```

```python
smo_status = pd.get_dummies(df_sp['smoking_status'], drop_first = True)
df_sp.drop(['smoking_status'], axis=1, inplace=True)
df_sp = pd.concat([df_sp, smo_status], axis=1)
```

```python
worktype = pd.get_dummies(df_sp['work_type'], drop_first = True)
df_sp.drop(['work_type'], axis=1, inplace=True)
df_sp = pd.concat([df_sp, worktype], axis=1)
```

```python
restype = pd.get_dummies(df_sp['Residence_type'], drop_first = True)
df_sp.drop(['Residence_type'], axis=1, inplace=True)
df_sp = pd.concat([df_sp, restype], axis=1)
```

```python
df_sp.rename(columns={'Yes':'Ever_Marr','Other':'Gender_Others'},inplace=True)
```

```python
df_sp.head()
```

```python
# Assigning values and splitting into training and test sets
X = df_sp.drop('stroke',axis=1).values
y = df_sp['stroke'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
# Applying feature scaling to the parameteres
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC 
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
models = [['Logistic Regresssion :',LogisticRegression()],
           ['Decision Tree Classification :', DecisionTreeClassifier()],
           ['RandomForest Classifier :',RandomForestClassifier(n_estimators = 20,criterion = 'entropy')],
           ['KNeighbours Classifier:', KNeighborsClassifier(n_neighbors = 5)],
           ['Gradient Boosting Classifier:', GradientBoostingClassifier(n_estimators = 999)],
           ['SVM :', SVC()],
           ['NB Classifier :', GaussianNB()]]
```

```python
for name_of_model,model_parms in models:
    curr_model = model_parms
    curr_model.fit(X_train, y_train)
    curr_predict = curr_model.predict(X_test)
    print(50*"=")
    print(20*" ",name_of_model,20*" ")
    print(50*"=")
    print("Accuracy Score  :", accuracy_score(y_test, curr_predict))
    print("Confusion Matrix :\n",confusion_matrix(y_test, curr_predict))
    print("Classification Report :\n", classification_report(y_test,curr_predict))
```
