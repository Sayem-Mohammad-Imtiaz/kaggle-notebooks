**SISSEJUHATUS**

Andmed on kogutud põhikooli matemaatika kursuse tudengitelt. Analüüsides käesolevaid andmeid, uurime, miks tarbivad õpilased alkoholi ja kuidas mõjutab alkoholi tarbimine õppeedukust.

* Suhted ja alkohol
* Vanus ja alkohol
* Õpilase suhe oma perekonnaga ja alkohol
* Elukoht ja alkohol
* Sugu ja alkohol
* Perekonnaseisu mõju alkoholi tarbimisele
* Kuidas ja kas alkohol mõjutab õppeedukust?
* KOKKUVÕTE

```python
import numpy as np
import pandas as pd 
%matplotlib inline


df = pd.read_csv("../input/student-mat.csv")
print("Kiire ülevaade õpilaste vanustest:")
print()
print("Keskmine vanus uuritavate seas on", df["age"].mean())
print("Väikseim vanus uuritavate seas on", df["age"].min())
print("Suurim vanus uuritavate seas on", df["age"].max())
```

```python
df.head(20)
```

Edasiseks kasutamiseks salvestan veeru "Weekalc", kus on õpilase nädalane alkoholitarbimine kokkuliidetud.

```python
df["Weekalc"] = df["Dalc"]+df["Walc"]
```

**Kas inimesed, kes on suhtes tarbivad vähem alkoholi?**

```python
df.groupby(["romantic"])["Weekalc"].mean().plot.pie(figsize=(6,6), colors=['b','y']);
```

Näeme, et suhtes oleks oluliselt ei mõjuta alkoholi tarbimist.

**Järgmiseks vaatame alkoholi tarbimist vastavalt vanusele.**

```python
df.groupby(["age"])["Weekalc"].mean().plot.bar(color = "b");
```

Näeme, et alates 20 aastaseks saamisest kasvab tarbimine oluliselt.

**Kuidas mõjutab õpilase suhe perekonnaga tarbitava alkoholi kogust?**

```python
df.groupby(["famrel"])["Weekalc"].mean().plot.bar(color = "brown");
```

Selgus, et suhe perekonnaga, mõjutab otseselt alkoholi tarbimist. Mida parem on perekondlik seis, seda vähem tarbitakse alkoholi.

**Vaatleme alkoholi tarbimise erinevust inimeste seas, kes elavad maal ja linnas.**

```python

df.groupby(["address"])["Weekalc"].mean().plot.bar(color = "g");
#linna_keskmine = a["U"]
#maa_keskmine = a["R"]

```

Üllatavalt paistab, et maakohtades elavad õpilased tarbivad keskmiselt pisut rohkem alkoholi.

**Vaatleme meeste ja naiste alkoholi tarbimis suhet.**

```python
soojaotus = df.groupby(["sex"])["Weekalc"].mean()
soojaotus.plot.pie(figsize=(6,6), colors=['r','g']);
```

Paistab, et meessoost õpilased tarbivad natukene rohkem alkoholi.

**Vanemate suhte(T tähendab vanemad elavad koos, A tähendab vanemad elavad lahus) pirukas.**

```python
df.groupby(["Pstatus"])["Weekalc"].mean().plot.pie(figsize=(6,6), colors=['r', 'b']);
```

Selgub, et vanemate suhte staatus, ei mõjuta oluliselt alkoholi tarbimist.

**Lõpuks vaatame ka, kas ja kuidas mõjutab alkoholi tarbimine eksaminehinde tulemusi. Koostan scatterploti:**

```python
df["Eksam"] = df["G3"]
df.plot.scatter("Weekalc", "Eksam", alpha=0.3);
```

Siit scatterplotist näeb selgeslt, et alkoholi tarbimisel on märgatav mõju õppeedukusel. Üldiselt, mida vähem õpilane joob, seda kõrgem on tal eksamitulemus.

**Kokkuvõtteks koostame grupeeringu, mis näiteab erinevaid andmeid sõltuvalt, kui palju õpilane nädalas alkoholi tarbib:**

```python
df.groupby("Weekalc")["absences","G3","health","famrel","goout"].mean().round(1)
```

```python
df.groupby("Weekalc")["absences","G3","health","famrel","goout"].mean().plot.bar(color = ["g","b","r","brown","y"]);
```
