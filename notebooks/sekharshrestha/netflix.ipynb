```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../input/netflix-shows/netflix_titles.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#To plot pie chart

counts = df['type'].value_counts().sort_index()
print(counts)
# Plot a pie chart
plt.figure(figsize = (10,7))
plt.pie(df['type'].value_counts(), labels = df['type'].value_counts().index, autopct="%.1f%%")

plt.show()
```

```python
#To plot pie chart

counts = df['rating'].value_counts().sort_index()
print(counts)
# Plot a pie chart
plt.figure(figsize = (15,10))
plt.pie(df['rating'].value_counts(), labels = df['rating'].value_counts().index, autopct="%.1f%%")

plt.show()
```

```python
#Top 20 categories in a col based on value_counts

plt.figure(figsize = (10,7))
chains = df['country'].value_counts()[:10] #top 10
sns.barplot(x = chains, y = chains.index)
plt.title("Countries")
plt.xlabel("Count")
```

```python
#Top 20 categories in a col based on value_counts

plt.figure(figsize = (10,7))
sns.barplot(x = df['release_year'].value_counts().index[:10], y = df['release_year'].value_counts()[:10])
plt.title("release_year")
plt.xlabel("release_year")
```

```python
#To see the bar graph of a col(never for continuous data)
plt.figure(figsize = (10,7))

sns.distplot(df['release_year'])
```

```python
#Top 20 categories in a col based on value_counts

plt.figure(figsize = (10,7))
chains = df['director'].value_counts()[:10] #top 10
sns.barplot(x = chains, y = chains.index)
plt.title("director")
plt.xlabel("Count")
```

```python
#Replace Seasons with '', and convert column to numeric

dftv.duration = dftv.duration.apply(lambda x: x.replace('Seasons',''))
dftv.duration = dftv.duration.apply(lambda x: x.replace('Season',''))

dftv['duration'] = dftv['duration'].apply(lambda x: float(x))
```

```python
dftv.head()
```

```python
#Replace mins with '', and convert column to numeric

dfm.duration = dfm.duration.apply(lambda x: x.replace('min',''))

dfm['duration'] = dfm['duration'].apply(lambda x: float(x))
```
