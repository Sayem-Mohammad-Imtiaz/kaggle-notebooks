# (3.4) Histogram, PDF, CDF

```python
# What about 1-D scatter plot using just one feature?
#1-D scatter plot of petal-length
import numpy as np
haberman_5yearsmore = haberman.loc[haberman["status"] == 1];
haberman_5yearsless = haberman.loc[haberman["status"] == 2];
plt.plot(haberman_5yearsmore["nodes"], np.zeros_like(haberman_5yearsmore['nodes']), 'o')
plt.plot(haberman_5yearsless["nodes"], np.zeros_like(haberman_5yearsless['nodes']), 'o')
plt.axis(aspect='equal');
plt.xlabel('Nodes')
plt.title('Nodes 1D ScatterPlot');
plt.show()
#Disadvantages of 1-D scatter plot: Very hard to make sense as points 
#are overlapping a lot.
#Are there better ways of visualizing 1-D scatter plots?
```

Here we can see short survival and long survival data is mostly overlapping.Therefore we will discard this graph.

```python
sns.FacetGrid(haberman, hue="status", size=8) \
   .map(sns.distplot, "nodes") \
   .add_legend();
plt.ylabel('Status Percentage');
plt.xlabel('Nodes');
plt.title('Nodes vs Status Percentage')
plt.show();
```

```python
sns.FacetGrid(haberman, hue="status", size=8) \
   .map(sns.distplot, "age") \
   .add_legend();

plt.ylabel('Status Percentage');
plt.xlabel('Age');
plt.title('Age vs Status Percentage')
plt.show();
```

```python
sns.FacetGrid(haberman, hue="status", size=8) \
   .map(sns.distplot, "year") \
   .add_legend();
plt.ylabel('Status Percentage');
plt.xlabel('Year');
plt.title('Nodes vs Status Percentage')
plt.show();
```

Observation: We cannot predict anything with these histograms as there is equal number of density in each data point.

```python
# Histograms and Probability Density Functions (PDF) using KDE
# How to compute PDFs using counts/frequencies of data points in each window.
# How window width effects the PDF plot.


# Interpreting a PDF:
## why is it called a density plot?
## Why is it called a probability plot?
## for each value of petal_length, what does the value on y-axis mean?
# Notice that we can write a simple if..else condition as if(petal_length) < 2.5 then flower type is setosa.
# Using just one feature, we can build a simple "model" suing if..else... statements.

# Disadv of PDF: Can we say what percentage of versicolor points have a petal_length of less than 5?

# Do some of these plots look like a bell-curve you studied in under-grad?
# Gaussian/Normal distribution.
# What is "normal" about normal distribution?
# e.g: Hieghts of male students in a class.
# One of the most frequent distributions in nature.
```

```python

# Need for Cumulative Distribution Function (CDF)
# How to construct a CDF?
# How to read a CDF?


counts, bin_edges = np.histogram(haberman_5yearsmore['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:], cdf)
plt.ylabel('Status Percentage for 5 years and more');
plt.xlabel('Nodes');
plt.title('Nodes vs Status Percentage')
plt.show();
```

From the above graph we can see that 85% of the cases of long survival if number of  nodes detected are less than 5

```python
# Need for Cumulative Distribution Function (CDF)
# We can visually see what percentage of versicolor flowers have a 
# petal_length of less than 1.6?
# How to construct a CDF?
# How to read a CDF?

#Plot CDF of nodes

counts, bin_edges = np.histogram(haberman_5yearsless['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.ylabel('Status Percentage for less than 5 years');
plt.xlabel('Nodes');
plt.title('Nodes vs Status Percentage')
plt.show();
```

Survival is nearly 55% of people who have nodes less than 5 and there are nearly 100% of people in short survival if nodes are > 40

# (3.5) Mean, Variance and Std-dev

```python
#Mean, Variance, Std-deviation,  
print("Means:")
print (np.mean(haberman_5yearsmore["nodes"]))
print (np.mean(haberman_5yearsless["nodes"]))

print("\nVariance:")
print (np.var(haberman_5yearsmore["nodes"]))
print (np.var(haberman_5yearsless["nodes"]))

print("\nStandard Deviation:")
print(np.mean(haberman_5yearsmore["nodes"]))
print(np.mean(haberman_5yearsless["nodes"]))



```

# (3.6) Median, Percentile, Quantile, IQR, MAD

```python
#Median, Quantiles, Percentiles, IQR.
print("\nMedians:")
print(np.median(haberman["nodes"]))
#Median with an outlier
print(np.median(np.append(haberman["nodes"],50)));
print(np.median(haberman_5yearsless["nodes"]))
print(np.median(haberman_5yearsmore["nodes"]))


print("\nQuantiles:")
print(np.percentile(haberman_5yearsless["nodes"],np.arange(0, 100, 25)))
print(np.percentile(haberman_5yearsmore["nodes"],np.arange(0, 100, 25)))

print("\n90th Percentiles:")
print(np.percentile(haberman_5yearsless["nodes"],90))
print(np.percentile(haberman_5yearsmore["nodes"],90))

from statsmodels import robust
print ("\nMedian Absolute Deviation")
print(robust.mad(haberman_5yearsless["nodes"]))
print(robust.mad(haberman_5yearsmore["nodes"]))
```

# (3.7) Box plot and Whiskers

```python

#Box-plot with whiskers: another method of visualizing the  1-D scatter plot more intuitivey.
# The Concept of median, percentile, quantile.
# How to draw the box in the box-plot?
# How to draw whiskers: [no standard way] Could use min and max or use other complex statistical techniques.
# IQR like idea.

#NOTE: IN the plot below, a technique call inter-quartile range is used in plotting the whiskers. 
#Whiskers in the plot below donot correposnd to the min and max values.

#Box-plot can be visualized as a PDF on the side-ways.

sns.boxplot(x='status',y='nodes', data=haberman)
plt.xlabel('Status Percentage');
plt.ylabel('Nodes');
plt.title('Nodes vs Status Percentile(Box Plot)')
plt.show()
```

In above box whiskers 25th percentile and 50th percentile are nearly same for 5 years and more and threshold for it is 0 to 7. Also, for 5 years less  there are 50th percentile of nodes are nearly same as long survive 75th percentile. Threshold for the 5 years less is 0 to 25 nodes and 75th% is 12 and 25th% is 1 or 2

# (3.8) Violin plots

```python
# A violin plot combines the benefits of the previous two plots 
#and simplifies them

# Denser regions of the data are fatter, and sparser ones thinner 
#in a violin plot

sns.violinplot(x="status", y="nodes", data=haberman, size=8)
plt.xlabel('Status Percentile');
plt.ylabel('Nodes');
plt.title('Nodes vs Status Percentile(Violin Plot)')
plt.show()
```
