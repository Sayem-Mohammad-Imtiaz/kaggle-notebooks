```python
from imblearn.pipeline import make_pipeline
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
smote=SMOTE()
knn_model=KNeighborsClassifier(n_neighbors=5)
svc_model=SVC(kernel='poly')
lr = LogisticRegression(solver='liblinear')
#x_s,y_s=smote.fit_resample(x,y)
new_pipe_knn=make_pipeline(trans_data,smote,knn_model)
new_pipe_svc=make_pipeline(trans_data,smote,svc_model)
new_pipe_lr=make_pipeline(trans_data,smote,lr)
```

```python
from sklearn.model_selection import StratifiedKFold
np.mean(cross_val_score(new_pipe_knn,x,y,cv=StratifiedKFold(n_splits=4),scoring='accuracy'))*100
```

```python
np.mean(cross_val_score(new_pipe_svc,x,y,cv=StratifiedKFold(n_splits=9),scoring='accuracy'))*100
```

```python
np.mean(cross_val_score(new_pipe_lr,x,y,cv=StratifiedKFold(n_splits=5),scoring='accuracy'))*100
```
