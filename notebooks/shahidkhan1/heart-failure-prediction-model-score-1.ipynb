# Import Libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

# Load Data

```python
df = pd.read_csv('../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv')
```

```python
df.head()
```

# Check Null Values

```python
df.isnull().sum()
```

There is no null values

```python
df.info()
```

There is no categorical values 

# Split The Data into Train and Test Set

```python
from sklearn.model_selection import train_test_split
y = df['DEATH_EVENT'] #dependent variable
x = df.drop('DEATH_EVENT', axis = 1) #Independent variable
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 101)
```

# Plot Pairplot to see which Model is Best for this 

```python
sns.pairplot(df, hue='DEATH_EVENT')
```

# Train the model with KNeighbors classifier

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X_train, y_train)
```

# Predict the Model

```python
y_predict = model.predict(X_test)
```

```python
y_predict
```

# Check the Score of the model

```python
model.score(X_test, y_predict)
```
