```python
msk = np.random.rand(len(df)) < 0.8
train = df[msk]
test = df[~msk]
```

```python
train.columns
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model
train_x = np.asanyarray(train[['X2 house age', 'X3 distance to the nearest MRT station',
       'X4 number of convenience stores', 'X5 latitude', 'X6 longitude'
       ]])
train_y = np.asanyarray(train[['Y house price of unit area']])

test_x = np.asanyarray(test[['X2 house age', 'X3 distance to the nearest MRT station',
       'X4 number of convenience stores', 'X5 latitude', 'X6 longitude']])
test_y = np.asanyarray(test[['Y house price of unit area']])


poly = PolynomialFeatures(degree=2)
train_x_poly = poly.fit_transform(train_x)
train_x_poly
```

```python
multi = linear_model.LinearRegression()
lr = multi.fit(train_x, train_y)
print ('Coefficients: ', lr.coef_)
print ('Intercept: ',lr.intercept_)
```

```python
from math import sqrt
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
preds_1 =lr.predict(test_x)
```

```python
Results = pd.DataFrame({'Algorithm': [],'MAE': [], 'R^2': [], "MSE":[], "RMSE":[]})
```

```python
res = pd.DataFrame({'Algorithm': ['Linear Regression'],'MAE': [mean_absolute_error(test_y, preds_1)], "R^2":[r2_score(test_y, preds_1)], "MSE":[mean_squared_error(test_y, preds_1)],"RMSE":[sqrt(mean_squared_error(test_y, preds_1))]})
 
Results = Results.append(res)
```

```python
clf = linear_model.LinearRegression()
train_y_ = clf.fit(train_x_poly, train_y)
# The coefficients
print ('Coefficients: ', clf.coef_)
print ('Intercept: ',clf.intercept_)
```

```python
test_x_poly = poly.fit_transform(test_x)
preds_2 = clf.predict(test_x_poly)
```

```python
res = pd.DataFrame({'Algorithm': ['Polynormial Regression'],'MAE': [mean_absolute_error(test_y, preds_2)], "R^2":[r2_score(test_y, preds_2)], "MSE":[mean_squared_error(test_y, preds_2)],"RMSE":[sqrt(mean_squared_error(test_y, preds_2))]})

Results = Results.append(res)
```
