```python
class_1=df.loc[df['Surv_Status']==1]
class_2=df.loc[df['Surv_Status']==2]
```

```python
# Age feature to class_1
counts, bin_edges=np.histogram(class_1['Age'],bins=10,density=True)
pdf=counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='pdf')
plt.plot(bin_edges[1:],cdf,label='cdf')
plt.grid()
plt.legend()
plt.title('CDF Plot 1: Age to class_1')
plt.show()
```
