```python
import seaborn as sb
dt=pd.read_csv('../input/ipl-complete-dataset-20082020/IPL Ball-by-Ball 2008-2020.csv')
dt.head(5)
```

```python
bat=dt.groupby('batsman').sum()['batsman_runs'].sort_values(ascending=False)[:10]
sb.barplot(bat.index,bat.values).set_title('Top 10 Batsmens with highest scores')
sb.set(rc={'figure.figsize':(11.7,8.27)})
```

```python
bol=dt.groupby('bowler').sum()['is_wicket'].sort_values(ascending=False)[:10]
sb.barplot(bol.index,bol.values).set_title('Top 10 Bowlers with highest number of wickets')
```

```python
six=dt[dt['batsman_runs']==6]
six_c=six.groupby(['batsman']).count()['batsman_runs'].sort_values(ascending=False)[:10]
sb.barplot(six_c.index,six_c.values).set_title('Top 10 Batsmans with highest number of 6s')
```

```python
four=dt[dt['batsman_runs']==4]
four_c=four.groupby('batsman').count()['batsman_runs'].sort_values(ascending=False)[:10]
sb.barplot(four_c.index,four_c.values).set_title('Top 10 Batsmans with highest number of 4s')
```

```python
cat=dt[(dt['dismissal_kind']=='caught')|(dt['dismissal_kind']=='caught and bowled')]
cat_s=cat.groupby('fielder').count()['is_wicket'].sort_values(ascending=False)[0:10]
sb.barplot(cat_s.index,cat_s.values).set_title('Top 10 Fielders with highest number of catches')
```

```python
cat=dt[dt['dismissal_kind']=='stumped']
cat_s=cat.groupby('fielder').count()['is_wicket'].sort_values(ascending=False)[0:10]
sb.barplot(cat_s.index,cat_s.values).set_title('Top 10 Wicket Keepers with highest number of Stumping')
```

```python
last_overs=dt[(dt['over']==20)|(dt['over']==18)|(dt['over']==19)]
last=last_overs.groupby('batsman').sum()['batsman_runs'].sort_values(ascending=False)[:10]
sb.barplot(last.index,last.values).set_title('Top 10 highest run scorers in the last 3 overs')
```

```python
out=dt.groupby('dismissal_kind').sum()['is_wicket'].sort_values(ascending=False)
sb.barplot(out.index,out.values).set_title('Number of Kind of dismissals')
```

```python
wik=dt.groupby('over').sum()['is_wicket'].sort_values(ascending=False)[:10]
sb.barplot(wik.index,wik.values).set_title('Overs in which Maximum wickets were taken')
```

```python
balls_faced=dt.groupby('batsman').count()['ball']
bat=dt.groupby('batsman').sum()['batsman_runs']
strike_r=bat/balls_faced
strike_r*=100
strike=strike_r.sort_values(ascending=False)[:10]
sb.barplot(strike.index,strike.values).set_title('Batsman with best Strike Rates')
```

```python
runs_conceded=dt.groupby('bowler').sum('total_runs')['total_runs']
balls_bowled=dt.groupby('bowler').count()['ball']
overs_bowled=balls_bowled/6
economy_rate=runs_conceded/overs_bowled
economy=economy_rate.sort_values(ascending=True)[:10]
sb.barplot(economy.index,economy.values).set_title('Bolwers with best economy rates')
```

```python
runs_conceded_last_overs=dt[(dt['over']==17) | (dt['over']==18) | (dt['over']==19)].groupby('bowler').sum('total_runs')['total_runs']
last_ball=dt[(dt['over']==17) | (dt['over']==18) | (dt['over']==19)].groupby('bowler').count()['ball']
last_overs=last_ball/6
economy_final_overs=runs_conceded_last_overs/last_overs
e=economy_final_overs.sort_values(ascending=True)[:10]
sb.barplot(e.index,e.values).set_title('Bolwers with best economy rates in 17th, 18th, and 19th over')
```

```python
batsman_runs_last_overs=dt[(dt['over']==17) | (dt['over']==18) | (dt['over']==19)].groupby('batsman').sum()['batsman_runs']
balls_faced_last_overs=dt[(dt['over']==17) | (dt['over']==18) | (dt['over']==19)].groupby('batsman').count()['ball']
strike_rate_last_overs=batsman_runs_last_overs/balls_faced_last_overs
strike_rate_last_overs*=100
strike=strike_rate_last_overs.sort_values(ascending=False)[:10]
sb.barplot(strike.index,strike.values).set_title('Batsman with best Strike Rates in the 17th,18th and 19th overs')
```
