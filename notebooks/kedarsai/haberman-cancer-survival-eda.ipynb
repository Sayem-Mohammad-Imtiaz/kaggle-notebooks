```python
import matplotlib.pyplot as plt
df=pd.read_csv(r'/kaggle/input/habermans-survival-data-set/haberman.csv',names=['Age','Op_Year','axil_nodes','Surv_status'])
```

```python
df.info()
```

```python
df.describe()
```

```python
df_1=df[df['Surv_status']==1].copy()
df_2=df[df['Surv_status']==2].copy()
```

## Univariate Analysis

```python
plt.figure(figsize=(10,5))
counts, bin_edges = np.histogram(df_1['Age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'r',label='AGE pdf_1')
plt.plot(bin_edges[1:], cdf,'b',label='AGE CDF_1')
plt.legend()
counts, bin_edges = np.histogram(df_2['Age'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'g',label='AGE pdf_2')
plt.plot(bin_edges[1:], cdf,'y',label='AGE CDF_2')
plt.legend()
plt.title('PDF, CDF of Age for Class 1 &2')
```

```python
plt.figure(figsize=(10,5))
counts, bin_edges = np.histogram(df_1['Op_Year'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'r',label='Op_Year PDF_1')
plt.plot(bin_edges[1:], cdf,'b',label='Op_Year CDF_1')
plt.legend()
counts, bin_edges = np.histogram(df_2['Op_Year'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'g',label='Op_Year PDF_2')
plt.plot(bin_edges[1:], cdf,'y',label='Op_Year CDF_2')
plt.legend()
plt.title('PDF, CDF of Op_Year for Class 1 &2')
```

```python
plt.figure(figsize=(10,5))
counts, bin_edges = np.histogram(df_1['axil_nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'r',label='axil_nodes PDF_1')
plt.plot(bin_edges[1:], cdf,'b',label='axil_nodes CDF_1')
plt.legend()
counts, bin_edges = np.histogram(df_2['axil_nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,'g',label='axil_nodes PDF_2')
plt.plot(bin_edges[1:], cdf,'y',label='axil_nodes CDF_2')
plt.legend()
plt.title('PDF, CDF of axil_nodes for Class 1 &2')
```

### Observation

There is 85% chance of survival after 5 years if a person has less than 10 auxilary lymph nodes
