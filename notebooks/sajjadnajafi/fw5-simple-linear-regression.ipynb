# Simple Linear Regression

In this homework we want to perform a simple linear regression on the "Real estate" dataset.

This dataset has 8 columns. The values of X1 to X6 columns affect the price per unit area of the house in the "Y house price of unit area" column.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('../input/real-estate-price-prediction/Real estate.csv')
```

## EDA:

```python
sns.pairplot(data = df,
             x_vars = ["X1 transaction date" ,
                      "X2 house age" ,
                      "X3 distance to the nearest MRT station",
                      "X4 number of convenience stores" ,
                      "X5 latitude" ,
                      "X6 longitude" ,
                      "Y house price of unit area"],
             y_vars = ["X1 transaction date" ,
                      "X2 house age" ,
                      "X3 distance to the nearest MRT station",
                      "X4 number of convenience stores" ,
                      "X5 latitude" ,
                      "X6 longitude" ,
                      "Y house price of unit area"]
            )
```

### Determine features and label:

```python
# Features:
X = df.drop(['Y house price of unit area'  , 'No'] , axis = 1)
# Label:
y = df['Y house price of unit area']
```

```python
X
```

```python
y
```

### Split dataset to train and test data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train.shape , X_test.shape
```

# Train the model

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
pd.DataFrame(model.coef_ , X.columns, columns=['Coefficient'])
```

## Predicting test data

```python
y_pred = model.predict(X_test)
```

## Evaluating The model

```python
from sklearn import metrics
MAE = metrics.mean_absolute_error(y_test, y_pred)
MSE = metrics.mean_squared_error(y_test, y_pred)
RMSE = np.sqrt(MSE)
```

```python
MAE , MSE , RMSE
```

```python
df['Y house price of unit area'].mean()
```

### Residuals

```python
test_residuals = y_test - y_pred
```

```python
sns.displot(test_residuals, bins = 25 , kde = True)
```
