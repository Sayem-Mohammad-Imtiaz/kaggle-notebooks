```python
synonyms = world_news[['flag','subject']]
synonyms
```

```python
import spacy
spacy.cli.download('en_core_web_md')
nlp = spacy.load("en_core_web_md")
```

```python
df = []
for k in synonyms:
    for w in synonyms[k]:
        df.append([w, k]+list(nlp(w).vector))
```

```python
from sklearn.linear_model import LogisticRegression

df = pd.DataFrame(df)
X = df[range(2, 302)]
y = df[1]


clf = LogisticRegression(random_state=0).fit(X, y)
```

```python
clf.score(X,y)
```
