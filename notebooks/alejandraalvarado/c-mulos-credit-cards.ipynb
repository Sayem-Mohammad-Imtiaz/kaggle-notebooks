```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
from sklearn.preprocessing import StandardScaler
#%% Leer la base
tarjetas = pd.read_csv('../input/ccdata/CC GENERAL.csv', index_col='CUST_ID')
tarjetas.columns
```

```python
#%%Análisis de cúmulos con 17 variables 
#X_3 = tarjetas.loc[:,['BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES']]

#Normalizando o re-escalando
transformer = StandardScaler()
X_17= pd.DataFrame(transformer.fit_transform(tarjetas))
#X_3_norm = pd.DataFrame(transformer.fit_transform(X_3))

#Sustituimos los "nan" por el promedio de la columna 14
promedio = X_17[14].mean()
X_17_norm = X_17.fillna(promedio)

#Encontrar número de cúmulos
#Gráfica de inercias
ks = range(1,25)
inertias = []
for k in ks:
    model = KMeans(n_clusters=k).fit(X_17_norm)
    inertias.append(model.inertia_)
plt.plot(ks, inertias, '-o')
plt.xlabel('Número de cúmulos, k')
plt.ylabel('inercia')
plt.xticks(ks)
plt.show()

#El máximo número de cúmulos para este ejecicio es de 5
model = KMeans(n_clusters=5).fit(X_17_norm)
```

```python
#%%#%%#GRUPO predicho para cada observación
tarjetas['Cluster'] = model.predict(X_17_norm)
X_17_norm['Cluster'] = model.predict(X_17_norm)
```

```python
#%%Gráficas con las variables más relevantes
# Graficando geografía vs día de la semana
sns.scatterplot(X_17_norm[0],X_17_norm[2],  hue='Cluster', data= X_17_norm)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X_17_norm[0],X_17_norm[2],X_17_norm[4],c=X_17_norm.Cluster)
plt.show()

plt.clf()
plt.scatter(X_17_norm[3],X_17_norm[4],c=X_17_norm.Cluster)
```
