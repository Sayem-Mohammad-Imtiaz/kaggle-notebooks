```python
def linelivechart(data,tr_1,tr_2):
    # prepare data frame
    df = data.iloc[:100,:]

    # import graph objects as "go"
    import plotly.graph_objs as go
    from plotly.offline import iplot

    # Creating trace1
    trace1 = go.Scatter(
                        x = df[tr_1],
                        y = df[tr_2],
                        mode = "lines",
                        name = "citations",
                        marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                        text= df.university_name)
    # Creating trace2
    trace2 = go.Scatter(
                        x = df.world_rank,
                        y = df.teaching,
                        mode = "lines+markers",
                        name = "teaching",
                        marker = dict(color = 'rgba(80, 26, 80, 0.8)'),
                        text= df.university_name)
    data = [trace1, trace2]
    layout = dict(title = 'Citation and Teaching vs World Rank of Top 100 Universities',
                  xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= True)
                 )
    fig = dict(data = data, layout = layout)
    iplot(fig)
```

```python
linelivechart(dataset,'world_rank', 'citations')
```

```python
 def scatterplot(data,header):
    # prepare data frames
    df2014 = data[ data.year == 2014].iloc[:100,:]
    df2015 = data[ data.year == 2015].iloc[:100,:]
    df2016 = data[ data.year == 2016].iloc[:100,:]
    # creating trace1
    import plotly.graph_objs as go
    from plotly.offline import iplot
    trace1 =go.Scatter(
                        x = df2014.world_rank,
                        y = df2014.citations,
                        mode = "markers",
                        name = "2014",
                        marker = dict(color = 'rgba(255, 128, 255, 0.8)'),
                        text= df2014.university_name)
    # creating trace2
    trace2 =go.Scatter(
                        x = df2015.world_rank,
                        y = df2015.citations,
                        mode = "markers",
                        name = "2015",
                        marker = dict(color = 'rgba(255, 128, 2, 0.8)'),
                        text= df2015.university_name)
    # creating trace3
    trace3 =go.Scatter(
                        x = df2016.world_rank,
                        y = df2016.citations,
                        mode = "markers",
                        name = "2016",
                        marker = dict(color = 'rgba(0, 255, 0.888, 0.8)'),
                        text= df2016.university_name)
    data = [trace1, trace2, trace3]
    layout = dict(title =header,
                  xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False),
                  yaxis= dict(title= 'Citation',ticklen= 5,zeroline= False)
                 )
    fig = dict(data = data, layout = layout)
    iplot(fig)
scatterplot(dataset,'Citation vs world rank of top 100 universities with 2014, 2015 and 2016 years')
```

```python
import plotly.graph_objs as go
from plotly.offline import iplot
def bar_chart(data,university_data):
    # prepare data frames
    df2014 = data[data.year == 2014].iloc[:5,:]

    x = df2014[university_data]

    trace1 = {
      'x': x,
      'y': df2014.citations,
      'name': 'citation',
      'type': 'bar'
    };
    trace2 = {
      'x': x,
      'y': df2014.teaching,
      'name': 'teaching',
      'type': 'bar'
    };
    data = [trace1, trace2];
    layout = {
      'xaxis': {'title': 'Top 5 universities'},
      'barmode': 'relative',
      'title': 'citations and teaching of top 5 universities in 2014'
    };
    fig = go.Figure(data = data, layout = layout)
    iplot(fig)
bar_chart(dataset, 'university_name')
```
