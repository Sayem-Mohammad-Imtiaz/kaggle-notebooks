```python
dataset= data.iloc[:,[14,13,17]].values
```

```python
X=dataset
```

```python
from sklearn.cluster import KMeans

def elbow_method(epsilon, figure= False):
    wcss = []


    diff = np.inf
    i = 1

    while diff > epsilon:
      print("iteracion No Clusters: k: {k}".format(k=i))



      kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
      kmeans.fit(X)


      if diff== np.inf:
          diff = kmeans.inertia_

      elif kmeans.inertia_== 0:
        wcss.append(kmeans.inertia_)
        break

      else:
        diff = (wcss [-1] - kmeans.inertia_)/wcss[-1]
      wcss.append(kmeans.inertia_)
      i += 1

    if figure:
      plt.plot(range(0,len(wcss)),wcss)
      plt.title('Metodo del Codo')
      plt.xlabel('numero de Cluster')
      plt.ylabel('WCSS')


    k = i-1
    return wcss,k
```

```python
epsilon= 0.05
wcss, _=elbow_method(epsilon, figure= True)
```

```python
epsilon = 0.33
_, k = elbow_method(epsilon, figure= False)
```

```python
Kmeans = KMeans(n_clusters = k, init = 'k-means++', max_iter = 300, n_init =  10, random_state = 0)
y_Kmeans = Kmeans.fit_predict(X)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter

ax.scatter3D(X[y_Kmeans == 0, 0], X[y_Kmeans == 0, 1], s = 100, c = 'blue', label = 'C0')
ax.scatter3D(X[y_Kmeans == 1, 0], X[y_Kmeans == 1, 1], s = 100, c = 'magenta', label = 'C1')
ax.scatter3D(X[y_Kmeans== 2, 0], X[y_Kmeans == 2, 1], s = 100, c = 'yellow', label = 'C2')
ax.scatter3D(X[y_Kmeans== 3, 0], X[y_Kmeans == 3, 1], s = 100, c = 'red', label = 'C2')

plt.title('Credit Card Cluster')
plt.xlabel('CREDIT_LIMIT')
plt.ylabel('PAYMENTS')
plt.legend()
plt.show()
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward')) 
plt.title('Dendrogram')
plt.xlabel('X1: Observations Number')
plt.ylabel('X2: Euclidean distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 3, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter

ax.scatter3D(X[y_hc == 0, 0], X[y_hc == 0, 1], s = 100, c = 'cyan', label = 'C0')
ax.scatter3D(X[y_hc == 1, 0], X[y_hc == 1, 1], s = 100, c = 'green', label = 'C1')
ax.scatter3D(X[y_hc == 2, 0], X[y_hc == 2, 1], s = 100, c = 'magenta', label = 'C2')

plt.title('Credit Card Cluster')
plt.xlabel('CREDIT_LIMIT')
plt.ylabel('PAYMENTS')
plt.legend()
plt.show()
```
