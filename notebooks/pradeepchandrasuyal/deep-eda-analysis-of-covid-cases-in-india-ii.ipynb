## we have already seen deep EDA analysis of covid data in the previous part, so now I will use geographical plotting for more better visualizations.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid_cases = pd.read_csv("../input/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv")
```

**choropleth**

Choropleth maps helps us to plot out information on a global or national wide. It allows us to see how a certain variable is distributed across a territory.


<img src="https://miro.medium.com/max/750/1*ZgMv41WkdwV6AmEU_FTVTw.gif">

```python
#importing Library
import plotly.express as px
import plotly.graph_objects as go
```

***for below plots you can simply point to the state to watch various feature properties***

```python
fig = px.choropleth(
    covid_cases,
    geojson="https://gist.githubusercontent.com/jbrobst/56c13bbbf9d97d187fea01ca62ea5112/raw/e388c4cae20aa53cb5090210a42ebb9b765c0a36/india_states.geojson",
    featureidkey='properties.ST_NM',
    locations='State/UTs',
    color='Total Cases',
    color_continuous_scale='earth',
    title = 'Total Covid Cases per State wise 2021'
)

fig.update_geos(fitbounds="locations", visible=False)
```
