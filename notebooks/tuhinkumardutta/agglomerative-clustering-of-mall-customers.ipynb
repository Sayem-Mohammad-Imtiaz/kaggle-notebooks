```python
import matplotlib.pyplot as plt, seaborn as sns
from scipy.cluster.hierarchy import linkage, dendrogram, cut_tree
```

```python
df = pd.read_csv('/kaggle/input/customer-segmentation-tutorial-in-python/Mall_Customers.csv')
df.head()
```

```python
# A copy of the dataframe is created
new = df.copy()
# Gender column in encoded
new.Gender = new.Gender.map(lambda x: 1 if x=='Male' else 0)
new.head()
```

```python
# CustomerID column is dropped
new = new.drop('CustomerID', axis=1)
new.info()
```

```python
# Columns are renamed
new = new.rename(columns = {'Annual Income (k$)':'annualIncome', 'Spending Score (1-100)':'spendingScore'})
new.head()
```

```python
# Dendrogram
links = linkage(new, method='complete')
dendrogram(links)
plt.show()
```

```python
df['cluster'] = cut_tree(Z=links, n_clusters=3)
df.head()
```

```python
# Boxplots created for cluster classification
for i in df.columns[2:-1]:
    sns.boxplot(x=df.cluster,y=df[i])
    plt.show()
```

```python
n=0.2
bar1= np.arange(2)
bar2=[i+n for i in bar1]
bar3=[i+n for i in bar2]

plt.figure(figsize=(20,8))
plt.bar(bar1,100*df[df.cluster==0].Gender.value_counts().sort_index()/df[df.cluster==0].Gender.value_counts().sum(), width=n, label='cluster 0')
plt.bar(bar2,100*df[df.cluster==1].Gender.value_counts().sort_index()/df[df.cluster==1].Gender.value_counts().sum(), width=n, label='cluster 1')
plt.bar(bar3,100*df[df.cluster==2].Gender.value_counts().sort_index()/df[df.cluster==2].Gender.value_counts().sum(), width=n, label='cluster 2')

plt.xticks(bar2, ['Female','Male'])
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Percentage Count', fontsize=12)
plt.title('Gender count in each cluster', fontsize=14)
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.legend()
plt.show()
```

## From the above plots it is clear that cluster 2 spends least in spite of having high annual income. Moreover, there are more males in cluster 2 than females unlike other clusters in which females are more.
