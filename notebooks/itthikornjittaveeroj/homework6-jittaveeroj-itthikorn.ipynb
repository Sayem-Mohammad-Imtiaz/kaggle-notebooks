```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

selected_country_1 = 'Canada'
selected_country_2 = 'Canada'

df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv', header = 0)
df1 = df1[df1['country']==selected_country_1]
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv', header = 0)
df2 = df2[df2['country']==selected_country_2]
df1.index = pd.to_datetime(df1['date'])
df2.index = pd.to_datetime(df2['date'])
print(df1)
print(df2)
```

In this exercise, I will be choosing Canada because I am interested in how the country is handling covid-19 recently. The Graph below shows the value of daily new cases in the country and the vaccination. The new daily cases are shown in red and vaccination in green.

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(df1.index, df1['daily_new_cases'], 'r')
ax1 = ax.twinx()
ax1.plot(df2.index, df2['daily_vaccinations'], 'g')
plt.show()
```

As you can see from the graph, the daily cases peaked around April this year, which is the thrid wave following May 2020 and in the new year. The cases have been decreasing since the vaccination started recently. According to the news article I read, 60% of the eligible Canadians have gotten the vaccine and that is why the new daily cases decrease.

# Improved Visualization

The following graph shows the same king of data as in the one above, but it is better visualized and can be zoomed in for more detailed information. 

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_new_cases= go.Scatter(x=df1.index, y=df1['daily_new_cases'], yaxis = 'y1', name = 'Daily confirmed cases')
daily_vaccinations = go.Scatter(x=df2.index, y=df2['daily_vaccinations'], yaxis = 'y2', name = 'Daily vaccinations')

layout_obj = go.Layout(title= 'COVID vs. Vaccines', xaxis=dict(title='Date'),yaxis=dict(title='Daily cases'),yaxis2=dict(title='Vaccinations', side = 'right', overlaying='y'))
fig = go.Figure(data=[daily_new_cases, daily_vaccinations],layout=layout_obj)
iplot(fig)
```

The code below lists all of the data of sum of daily new cases and the people vaccinated per hundred. For Canada, the sum of all daily cases is 1359172 people and the number of people vaccinated per hundred is 4.49. The rate of vaccination of this country is better than many countries like Japan, Vietnam or Thailand, but it is still not that much compared to countries like the UK, Germany, Israel,and the USA.

```python
df1= pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv', header = 0)
df1 = df1.groupby('country').sum()['daily_new_cases'].sort_values(ascending = False)
print(df1.to_string())
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv', header = 0)
df2 = df2.groupby('country').max()['people_fully_vaccinated_per_hundred'].sort_values(ascending = True)
print(df2.to_string())
```
