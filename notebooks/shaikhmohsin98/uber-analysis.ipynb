```python
Category =pd.crosstab(index=uber_data['CATEGORY*'],columns='Count of travels as per category')

Category.plot(kind='bar',color='b',alpha=0.8)
plt.legend()
Category
```

```python
plt.figure(figsize=(15,15))
pd.Series(uber_data['STOP*']).value_counts()[:25].plot(kind = "pie")
plt.title("Cab Rides STOP Location frequency")
plt.xticks(rotation = 50)
```

# **Profit Earned by Miles**

```python
a = sb.factorplot(x="PURPOSE*", y="MILES*", hue="CATEGORY*", data=uber_data,
                   size=15, kind="bar", palette="muted")
#from the graph is clearly seen that the main contributors for miles are:
#in Business category: meetings and customer; private: commute and charity 
a.fig.suptitle('Miles was earned per category and purpose', fontsize= 25)
a.fig.set_size_inches(15, 15)
a.set_xlabels('PURPOSE*', fontsize= 20)
a.set_ylabels('MILES*', fontsize= 20)
```

# **Rides Frequently Used**

```python
plt.figure(figsize=(12,7))
sb.countplot(uber_data['MILES*'], order=uber_data['MILES*'].value_counts().index, palette='deep')
plt.axhline(uber_data['MILES*'].value_counts().mean(),linestyle='dotted', color = 'darkblue', label='Mean Trips across Months')
plt.legend()
plt.show()
```

Here we can conclude that each cab travels by 3-4 miles away from each distance counts. The mileage of each is 30-40 minutes near by the particular destination.
