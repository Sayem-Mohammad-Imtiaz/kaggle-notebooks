The happiness of a person can help measure the well-being and contentment of a person. There are a lot of factors that can affect a person's happiness. Some of these are as follows: economy of their country, family, health, freedom, government corruption, generosity, and dystopia as seen in the chosen data set.

1. Which factor mostly affects the happiness rate of a country positively? Economy
2. Which factor mostly affects the happiness rate of a country negatively? None
3. Can happiness affect the economy of a country? Yes.

The factors don't affect the happiness score negatively, they also don't have strong correlation with each other. The economy is affected by the productivity of the people, the happier they are, the possibility of citizens being more productive is higher. Citizens are also happier if their country's economy is high because they believe their country is developed. Next is health, there's lesser stress among the countrymen because they are happy, thus their lifespan expans.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds = pd.read_csv("../input/world-happiness/2016.csv")
```

Scatter Plot of Economy and Happiness

```python
ds.plot.scatter(x = 'Economy (GDP per Capita)', y = 'Happiness Score')
```

Correlation of Economy and Happiness

```python
samp = {'Happiness': ds['Happiness Score'], 'Economy': ds['Economy (GDP per Capita)']}
dsamp = pd.DataFrame(samp)
dsamp.corr('pearson')
```
