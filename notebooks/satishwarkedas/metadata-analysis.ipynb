1. Remove rows with no competitionid, no reward
2. Identify the submission corresponding to only "featured" and "research" competitions
3. Based on the EvaluationAlgorithmIsMax -> multiple scores with -1 (for FALSE), 1 (for TRUE)
4. Prepare a column of dScore - PublicScore - PrivateScore



```python
def convert_score(row):
    if row['EvaluationAlgorithmIsMax'] == True:
        return -1
    elif row['EvaluationAlgorithmIsMax'] == False:
        return 1
    else:
        return 0
dfSub['ConversionFactor'] = dfSub.apply(lambda row: convert_score(row), axis = 1)
```

```python
dfSub = dfSub.reset_index()
# dfSub = dfSub.set_index('Id')
dfSub.dropna(subset = ['CompetitionId'], inplace = True)
print(dfSub.shape)
# dfSub.dropna(subset = ['RewardQuantity'], inplace = True)
print(dfSub.shape)
dfSub = dfSub[dfSub['CompetitionType'].isin(['Featured', 'Research', 'GE Quests', 'Prospect'])]
dfSub = dfSub[dfSub['ConversionFactor'] != 0]
dfSub['PublicScoreFullPrecisionConverted'] = dfSub['PublicScoreFullPrecision'] * dfSub['ConversionFactor']
dfSub['PrivateScoreFullPrecisionConverted'] = dfSub['PrivateScoreFullPrecision'] * dfSub['ConversionFactor']
dfSub['ScoreDifferential'] = dfSub['PublicScoreFullPrecision'] - dfSub['PrivateScoreFullPrecision']
```

4. Calculate the standard deviation of this difference for each team

```python
df = dfSub.groupby(['TeamId'])['ScoreDifferential'].agg(['mean', np.std, 'count'])
df = df.reset_index()
print(df.head())
dfSub2 = pd.merge(dfSub, df, on = ['TeamId'])
dfSub2['ScoreDifferentialStandardized'] = dfSub2['ScoreDifferential'] - dfSub2['mean']
dfSub2['ScoreDifferentialStandardized'] = dfSub2['ScoreDifferentialStandardized'] / dfSub2['std']
dfSub2['ScoreDifferentialNonStandardized'] = dfSub2['ScoreDifferential'] / dfSub2['std']
del dfSub, df
```

5. Create other rows for each submission -> score differential/standard deviation, whether team or not, team size, feedback earlier received
5.1 Check if submission date and score date are too different

```python
dfSub2['SubmissionDateNew'] = pd.to_datetime(dfSub2['SubmissionDate'], format = '%m/%d/%Y')
dfSub2['ScoreDateNew'] = pd.to_datetime(dfSub2['ScoreDate'], format = '%m/%d/%Y')
dfSub2['FeedbackDelay'] = (dfSub2['ScoreDateNew'] - dfSub2['SubmissionDateNew'])/np.timedelta64(1,'D')
dfSub2 = dfSub2.sort_values(by = ['CompetitionId', 'TeamId', 'ScoreDateNew'])
# print(dfSub2.head())
# dfSub2.dropna(subset = ['ScoreDate'], inplace = True)
dfSub2['ScoreDateOld'] = dfSub2['ScoreDateNew'].shift(1)
dfSub2['TimefromEnabledDate'] = (dfSub2['ScoreDateNew'] - dfSub2['EnabledDateNew'])/np.timedelta64(1,'D')
dfSub2['TimeinCompetition'] = dfSub2['TimefromEnabledDate']/dfSub2['CompetitionDuration']
dfSub2['TimeuntilDeadlineDate'] = (dfSub2['DeadlineDateNew'] - dfSub2['ScoreDateNew'])/np.timedelta64(1,'D')
dfSub2['TimeLeftinCompetition'] = dfSub2['TimeuntilDeadlineDate']/dfSub2['CompetitionDuration']
```

6. make team id the index
7. calculate the difference between score dates
8. calculate the number of feedback prior to the current feedback
9. drop all unnecessary columns
10. Score differential calculations ->

```python
dfSub2['TimefromLastFeedback'] = dfSub2['ScoreDateNew'] - dfSub2['ScoreDateOld']
dfSub2['TimefromLastFeedback'] = dfSub2['TimefromLastFeedback']/np.timedelta64(1,'D')
dfSub2 = dfSub2.sort_values(by = ['CompetitionId', 'TeamId', 'SubmissionDateNew'])
dfSub2['SubmissionDateOld'] = dfSub2['SubmissionDateNew'].shift(1)
dfSub2['TimefromLastSubmission'] = dfSub2['SubmissionDateNew'] - dfSub2['SubmissionDateOld']
dfSub2['TimefromLastSubmission'] = dfSub2['TimefromLastSubmission']/np.timedelta64(1,'D')
dfSub2['RewardPresent'] = ~dfSub2['RewardQuantity'].isna()
```

10. merge with team size/ team presence to get the team based association

```python
def label_score(row):
    if row['ScoreDifferential'] > 0:
        return 1
    elif row['ScoreDifferential'] == 0:
        return 0
    elif row['ScoreDifferential'] < 0 :
        return -1
dfSub2['DifferentialLabel'] = dfSub2.apply(lambda row: label_score(row), axis=1)
```
