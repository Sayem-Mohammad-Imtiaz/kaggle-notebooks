# Checking Missing Values

```python
df.isnull().sum()
```

**The Dataset contains -1, 0 and 1 values in column entries, so I will be changing them to NaN**

```python
df.replace(-1, np.nan, inplace = True)
df.replace('-1', np.nan, inplace = True)
df.replace(-1.0, np.nan, inplace = True)
```

```python
plt.figure(figsize=(10,5))
colormap = plt.cm.plasma
sns.heatmap(df.corr(), annot=True, cmap=colormap)
```

**Now Let us again look for missing values**

```python
missing_df = df.isnull().sum()
missing_df
```

**So, after we modified the data we see that compared to only 1 missing values, here there are many missing values**

# Assessing how much missing data is present

```python
number_of_rec = df.shape[0]
print(number_of_rec)

prop_msval = pd.DataFrame(missing_df, columns = ['Missing after update'])
prop_msval = prop_msval/(number_of_rec/100)
display(prop_msval)
```

**Easy Apply column has 96.4% missing values and so the column will not be of any help to us, Competitors also has 76% missing values, so we will have to drop it as well**

# Displaying Missing Values

```python
prop_msval.plot(kind='barh', figsize=(15,10))
plt.xlabel('Missing data frequency')
```

**So as is evident from above we can remove these columns that have more missing values, as we cannot get any useful information out of them now**

**Let's start with analyzing columns that have missing values of more than 10%**

```python
prop_msval2 = prop_msval[prop_msval['Missing after update']>10.0]
display(prop_msval2)
```

**Columns that have missing values of more than 20% will be dropped**

```python
to_drop = prop_msval[prop_msval['Missing after update']>20.0]
to_drop = to_drop.transpose()
display(to_drop)
```

```python
df = df.drop(to_drop,axis=1)
df = df.drop('Unnamed: 0',axis=1)
df.head(5)
```
