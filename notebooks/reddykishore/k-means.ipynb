```python
df.drop('Gender',inplace=True,axis=1)
```

```python
from plotly.subplots import make_subplots
l=[]
m=[]
n=range(1,11)
row=0
col=0
fig = make_subplots(rows=2, cols=2)
for i in n:
    
    clf = KMeans(n_clusters=i)
    clf.fit(df)
    if(i<6):
        y_means = clf.predict(df)
        y_means=pd.DataFrame({'output':y_means})
        data=pd.concat([df,y_means],axis=1)
        plt.rcParams['figure.figsize'] = (28, 10)
        sns.set(style = 'whitegrid')
        
        fig = px.scatter_3d(data,x='Age', y='Annual Income (k$)', z='Spending Score (1-100)',title='cluster'+str(i),
                            
                    color='output',width=1200, height=800)
       
        if(i%2==0):
            col+=1
        else:
            row+=1
        m.append(fig)
    l.append(clf.inertia_)

```

```python
from plotly.subplots import make_subplots
for i in m:
    i.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 8)
plt.plot(n,l)
```
