```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import missingno as msno
```

```python
df = pd.read_csv('../input/pokemon/Pokemon.csv')
```

```python
msno.matrix(df, figsize = (10,5))
```

```python
df.isnull().sum()
```

```python
df['Type 2'] = df['Type 2'].fillna('None')
```

```python
len(df)
```

The are 800 pokemons in the dataset

```python
sns.countplot(data = df, x = 'Legendary')
plt.ylabel('Number of pokemons')
plt.title('Number of common and legendary pokemons')
plt.show()
```

```python
sns.countplot(data = df, x = 'Generation', hue = 'Legendary')
plt.ylabel('Number of pokemons')
plt.title('Number of common per generation')
plt.show()
```

```python
ordem = df['Type 1'].value_counts()
ordem = pd.DataFrame(ordem)
sorted(ordem)
sns.countplot(data = df, x = 'Type 1', order = ordem.index)
plt.ylabel('Number of pokemons')
plt.title('Number of common per Type 1')
plt.xticks(rotation = 45, horizontalalignment = 'right')
plt.show()
```

The most common Type 1 is Water, Normal, Grass, Bug and Psychic.

```python
ordem = df['Type 2'].value_counts()
ordem = pd.DataFrame(ordem)
ordem = ordem.reset_index()
sorted(ordem)
sns.countplot(data = df, x = 'Type 2', order = ordem.loc[ordem['index'] != 'None']['index'])
plt.ylabel('Number of pokemons')
plt.title('Number of common per Type 2')
plt.xticks(rotation = 45, horizontalalignment = 'right')
plt.show()
```

The most common Type 2 is Flying, Ground, Poison, Psychic and Fighting.

```python
fig, ax = plt.subplots(figsize = (20,10))
ordem = df.pivot_table(values = 'Total', columns = 'Type 1', aggfunc = 'median')
ordem = ordem.T
ordem = ordem.reset_index(inplace = False)
ordem =ordem.sort_values(by = 'Total', ascending = False)
sns.boxplot(data = df, x = 'Type 1', y = 'Total', order = ordem['Type 1'], palette = 'summer_r')
plt.xlabel('Total stats of pokemons')
plt.title('Total stats distribution of pokemons by main type (type 1)')
plt.show()
```

Considering the Type 1 only, if we order the plot by the median of the column Total, the strongest Type 1 is DRAGON

```python
fig, ax = plt.subplots(figsize = (20,10))
ordem = df.pivot_table(values = 'Total', columns = 'Type 2', aggfunc = 'median')
ordem = ordem.T
ordem = ordem.reset_index(inplace = False)
ordem = ordem.loc[ordem['Type 2'] != 'None']
ordem = ordem.sort_values('Total', ascending = False)
sns.boxplot(data = df, x = 'Type 2', y = 'Total', palette = 'summer_r', order = ordem['Type 2'])
plt.xlabel('Type 2')
plt.title('Total stat distribution by type 2')
plt.show()
```

Ploting the boxplot of Total stat per Type 2 and classifying it by median of Total Stat we also see that DRAGON type is once again the the podium

```python
fig, ax = plt.subplots(figsize = (16,8))
sns.kdeplot(data = df, x = 'Total', hue = 'Legendary', cumulative = False)
```

Here we can see the total stat distribution among legendary and non-legendary pokemons. Legendary pokemons tend to be way more powerful.

```python
fig, ax = plt.subplots(figsize = (18,9))
ordem = df.query('Legendary == 1')
ordem = ordem.pivot_table(values = 'Total', columns = 'Type 1', aggfunc = 'median')
ordem = ordem.T
ordem = ordem.reset_index(inplace = False)
ordem = ordem.sort_values(by = 'Total', ascending = False)
sns.boxplot(data = df.query('Legendary == 1'), x = 'Type 1', y = 'Total', order = ordem['Type 1'])
```

When we take a look only to the legendaries, Dragon Type 1 is in the second place

```python
fig, ax = plt.subplots(figsize = (10,5))
sns.boxplot(data = df, x = 'Generation', y = 'Total', palette = 'RdBu')
```

The strongest generation is the fourth

```python
stats = ['HP','Attack','Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
```

```python
sns.pairplot(data = df[stats], corner = True)
```

No clear relationship between stats list

```python
fig, ax = plt.subplots(figsize = (12,8))
sns.heatmap(df[stats].corr(), annot = True, cmap = sns.color_palette("Blues", as_cmap=True), mask = np.triu(np.ones_like(df[stats].corr())), square = True, vmax = 1, center = 0)
plt.ylim(0,6)
plt.show()
```

Other way to see the correlation is to use the method pd.DataFrame.corr(). Confirming that there is no strong correlation between the features
