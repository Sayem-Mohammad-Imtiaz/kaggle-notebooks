```python
#Covid 19 Visualization

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
#This includes analysis of country wise and day wise covid 19 active cases, recovered, deaths and total number of cases.
```

```python
#Reading the necessary data and exploring the first five dataset values

country_wise = pd.read_csv("../input/corona-virus-report/country_wise_latest.csv")
day_wise = pd.read_csv("../input/corona-virus-report/day_wise.csv")
country_wise.head(1000)
```

```python
day_wise.head(1000)
```

```python

import plotly.io as pio
pio.renderers.default = "svg"


import plotly.offline as pl
import plotly.graph_objs as go
import plotly.express as px
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)

import warnings
warnings.filterwarnings(action='ignore')

```

```python
trace1=go.Scatter(x=day_wise.Date,y=day_wise.Confirmed,mode='lines',name='Confirmed',text='Case Confirmed.')

trace2=go.Scatter(x=day_wise.Date,y=day_wise.Deaths,mode='lines',name='Deaths',text='Deaths caused by virus')
data1=[trace1,trace2]
layout=dict(title='Confirmed Case',xaxis=dict(title='Days'),yaxis=dict(title='Number of Case and Deaths'))
fig=dict(data=data1,layout=layout)
iplot(fig)
```

```python
fig1=px.line(day_wise,x='Date',y='Confirmed')
iplot(fig1)
```

```python
country_wise=country_wise.sort_values(by='Confirmed',ascending=False)
country_wise.head(10)
country_wise.rename(columns={'Country/Region':'Country'},inplace=True)
country_wise.head(10)

```

```python
trace1=go.Bar(x=country_wise.Country[:15],y=country_wise.Confirmed[:15],name='Confirmed')

data=[trace1]

layout=dict(title='Countries Confirmed Cases',xaxis=dict(title='countries'),yaxis=dict(title='no. of cases'))

fig=dict(data=data,layout=layout)
iplot(fig)
```

```python
bar_graph=px.bar(country_wise[:15],x='Country',y='Confirmed',color='Deaths / 100 Cases')
iplot(bar_graph)
```

```python
project=px.scatter_geo(country_wise,locations='Country',locationmode='country names',color='WHO Region',hover_name="Country",size='Confirmed',projection="kavrayskiy7",title='No. Of Confirmed Cases By Countries')
project.show()
```
