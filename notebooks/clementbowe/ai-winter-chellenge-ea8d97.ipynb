```python
with open("../input/joketext-files-fixed/shortjokesfixed.txt") as sf:
    shortjokestext = sf.read()
    
with open("../input/joketext-files-fixed/qajokesfixed.txt") as qf:
    qajokestext = qf.read()
```

```python
with open("../input/extrajoke-textfiles2/redditjokes.txt") as rf:
    redditjokestext = rf.read()
    
with open("../input/extrajoke-textfiles2/wockajokes.txt") as wf:
    wockajokestext = wf.read()
```

```python
shortjokesmodel = markovify.NewlineText(shortjokestext)
```

```python
qajokesmodel = markovify.NewlineText(qajokestext)
```

```python
redditjokesmodel = markovify.NewlineText(redditjokestext)
```

```python
wockajokesmodel = markovify.NewlineText(wockajokestext)
```

```python
combo_models = markovify.combine([ shortjokesmodel, qajokesmodel, redditjokesmodel, wockajokesmodel ], [ 1, 0.5, 0.65, 1 ])
```

```python
combo_models.compile(inplace=True)
```

```python
#Function to control what is "generated" by the AI
for i in range(10):
    genJoke = combo_models.make_sentence(tries=100)
    print(genJoke)
    print("")
```
