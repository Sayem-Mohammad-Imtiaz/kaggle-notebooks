**Q1: Create a faceted plot in ggplot of a public data set use size, shape and color as well as facets.**

```python
# data cleaning
chocolate['Cocoa_Percent_num'] = chocolate['Cocoa\nPercent'].str.replace('\%','').astype(float)
chocolate['ifUSA'] = np.nan
chocolate.loc[chocolate['Company\nLocation'].str.contains('U.S.A.'),'ifUSA'] = 'U.S.A.'
chocolate.loc[~chocolate['Company\nLocation'].str.contains('U.S.A.'),'ifUSA'] = 'Other Countries'
chocolate.ifUSA.astype('category')
chocolate.head()
```

```python
# ggplot
ggplot(chocolate, aes(x='Cocoa_Percent_num', y='Rating', color='ifUSA', shape='ifUSA', size='REF'))+\
geom_point()+\
facet_wrap('Review\nDate')+\
ggtitle('Cocoa Percent VS. Rating')+\
labs(x='Cocoa Percent(%)', y='Rating')
```

**Q2: Create a Correlation Heatmap in Seaborn using a public dataset.**

```python
plt.subplots(figsize=(20,15))
plt.axes().set_title("Chocolate Bar Rating Correlation Heatmap")
corr = chocolate.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
```
