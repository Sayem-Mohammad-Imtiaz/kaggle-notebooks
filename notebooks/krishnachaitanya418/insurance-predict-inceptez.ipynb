```python
data = pd.read_csv('/kaggle/input/insurance-premium-prediction/insurance.csv')
```

```python
data.head()
```

```python
data.corr()
```

```python
data['smoker'].value_counts()
```

```python
data['smoker'].replace({'no':0,
                         'yes':1 },inplace=True)
```

```python
data['smoker'].value_counts()
```

```python
data['sex'].value_counts()
```

```python
data['sex'].replace({'male':0,
                    'female':1 },inplace=True)
```

```python
data['sex'].value_counts()
```

```python
data['region'].value_counts()
```

```python
data['region'].replace({'southeast':0,
                        'northwest':1,
                        'southwest':2,
                        'northeast':3},inplace=True)
```

```python
data['region'].value_counts()
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scale=StandardScaler()
num_data = data.drop(columns = ['sex','smoker','region'])
```

```python
num_data.head()
```

```python
data_std = std_scale.fit_transform(num_data)
```

```python
num_preprocess = pd.DataFrame(data_std,columns = num_data.columns)
```

```python
 num_preprocess
```

```python
data_preprosses = pd.concat( [num_preprocess,data['sex'],data['smoker'],data['region']],axis=1)
```

```python
data_preprosses.head()
```

```python
data_preprosses.corr()
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
X = data_preprosses.drop('expenses',axis='columns')
```

```python
y = data_preprosses['expenses']
```

```python
model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
data_predict = model.predict(X)
```

```python
data_predict
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y,data_predict) #MSE
```

```python
np.sqrt(mean_squared_error(y,data_predict)) #RMSE
```

```python
from sklearn.model_selection import train_test_split, KFold, cross_val_score
```

```python
np.abs(cross_val_score(model, X, y, scoring = 'neg_root_mean_squared_error', cv = 10)) #K Fold (10)
```

```python
np.mean(np.abs(cross_val_score(model, X, y, scoring = 'neg_root_mean_squared_error', cv = 10)))
```

```python
X_train, X_test, y_train, y_test  = train_test_split(X,y,test_size = 0.3, random_state=90)
```

```python
model.fit(X_train,y_train)
```

```python
train_pred = model.predict(X_train)
test_pred = model.predict(X_test)
```

```python
np.sqrt(mean_squared_error(y_test, test_pred))
```

```python
np.sqrt(mean_squared_error(y_train,train_pred))
```
