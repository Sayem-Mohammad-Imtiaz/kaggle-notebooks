**Dataset description**

* The dataset used is Pima Indian Diabetes Dataset from UCI Machine Learning Repository.
* Each entry in this dataset represents a female at least 21 years old of Pima Indian heritage.
* There are 768 entries (patients) and each entry has 8 features and 1 target (class label).

The features (columns) are:
1. Number of times pregnant.
2. Plasma glucose concentration.
3. Diastolic blood presure in mm.
4. Triceps skin fold thickness in mm.
5. 2-hour serun insulin in mu U/ml.
6. Body mass index.
7. Diabetes pedigree function.
8. Age in years.
