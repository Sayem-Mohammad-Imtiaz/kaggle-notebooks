# 2. Data structure- Dataframe
DataFrame is a 2-dimensional labeled data structure with columns of potentially different types. You can think of it like a spreadsheet or SQL table, or a dict of Series objects. It is generally the most commonly used pandas object.

```python
# Create dataframe from series
record1=pd.Series({'name':'a','class':'physics','score':85})
record2=pd.Series({'name':'b','class':'chemistry','score':82})
record3=pd.Series({'name':'c','class':'biology','score':90})
df=pd.DataFrame([record1,record2,record3],index=['school1','school2','school3'])
df.head()
```

```python
#Create dataframe from dictionary
students=[{'name':'a','class':'physics','score':85},
          {'name':'b','class':'chemistry','score':82},
          {'name':'c','class':'biology','score':90}]
    
df=pd.DataFrame(students,index=['school1','school2','school1'])
df.head()
```

### Subsetting Rows

```python
#select data based on multipul axes using .loc. It takes two parameters, the row index and the column names.
df.loc['school1','name']
```

# 5. Merging Dataframe

```python
staff_df=pd.DataFrame([{'name':'Kelly','roll':'Director'},
                      {'name':'Sally','roll':'HR'},
                      {'name':'James','roll':'Manager'}])
staff_df=staff_df.set_index('name')
student_df=pd.DataFrame([{'name':'James','school':'Business'},
                      {'name':'Mike','school':'Law'},
                      {'name':'Sally','school':'Engineering'}])
student_df=student_df.set_index('name')
print(staff_df)
print(student_df)
```

### 1) using index to merge dataframe

```python
pd.merge(staff_df,student_df, how='outer',left_index=True,right_index=True)
```

```python
pd.merge(staff_df,student_df, how='inner',left_index=True,right_index=True)
```

```python
#get a list of all staff 
pd.merge(staff_df,student_df, how='left',left_index=True,right_index=True)
```

```python
#get a list of all students
pd.merge(staff_df,student_df, how='right',left_index=True,right_index=True)
```

### 2) using columns to merge dataframe

```python
staff_df=staff_df.reset_index()
student_df=student_df.reset_index()

pd.merge(staff_df,student_df, how='right',on='name')
```

```python
staff_df=pd.DataFrame([{'name':'Kelly','roll':'Director','location':'NY'},
                      {'name':'Sally','roll':'HR','location':'TX'},
                      {'name':'James','roll':'Manager','location':'FL'}])

student_df=pd.DataFrame([{'name':'James','school':'Business','location':'1024 Avenue'},
                      {'name':'Mike','school':'Law','location':'2733 Heart street'},
                      {'name':'Sally','school':'Engineering','location':'512 wilson'}])

pd.merge(staff_df,student_df, how='right',on='name')
#The merge function append _x and _y to help differentiate the columns with same column name
```

```python
staff_df=pd.DataFrame([{'first name':'Kelly','last name':'Des','roll':'Director','location':'NY'},
                      {'first name':'Sally','last name':'Br','roll':'HR','location':'TX'},
                      {'first name':'James','last name':'Wilde','roll':'Manager','location':'FL'}])

student_df=pd.DataFrame([{'first name':'James','last name':'Ham','school':'Business','location':'1024 Avenue'},
                      {'first name':'Mike','last name':'Smith','school':'Law','location':'2733 Heart street'},
                      {'first name':'Sally','last name':'Br','school':'Engineering','location':'512 wilson'}])

pd.merge(staff_df,student_df, how='inner',on=['first name','last name'])
```

## merge dataframe vertically

```python
pd.concat([staff_df,student_df], join ='outer')
```

```python
pd.concat([staff_df,student_df], join ='inner')
```

```python
pd.concat([staff_df,student_df], keys=['staff','student'])
```

# 6. Group by Function

```python
df = pd.read_csv('../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
df.head()
```

```python
df.groupby('Year').agg(np.average)
```

```python
df.groupby('Year').agg({'User Rating':np.average})
```

```python
df.groupby('Year')['User Rating'].agg(np.average)
```

```python
df.groupby('Year').agg({'Price':(np.average, np.sum)})
```

```python
df.groupby(['Year','User Rating']).agg({'Price':(np.average, np.sum)})
```
