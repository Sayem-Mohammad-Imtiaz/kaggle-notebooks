* We need to make an adjustment on the Premiere feature,it should be datetime object, and also transform it in order to consider it in more detail.



```python
df['Premiere'] = pd.to_datetime(df['Premiere'])
```

```python
df['Year'] = df['Premiere'].apply(lambda x:x.year)
df['Day'] = df['Premiere'].apply(lambda x:x.day)
months ={1: 'January',2: 'Febuary',3: 'March',4: 'April',5: 'May',6: 'June',7: 'July',8: 'August',9: 'September',10: 'October',
        11: 'November',12: 'December'}
df['Month'] = df['Premiere'].apply(lambda x:months[x.month])
```

```python
df = df.drop('Premiere',axis=1)
df.head()
```

```python
df.groupby('Year').count()['Title'].plot(kind='bar')
plt.title('Movie Releases Per Year')
plt.xlabel('Year')
plt.ylabel('Total Number')
```

```python
df.groupby('Month').count()['Title'].plot(kind='bar')
plt.title('Movie releases per month')
plt.xlabel('Months')
plt.ylabel('Movie count')
```

```python
sns.set(rc={'figure.figsize':(14, 8)})
ax = sns.countplot(df.Month, hue= df.Year, palette='bright',data = df)
ax.set_title('Number of new movies by month')
plt.xlabel('Month')
plt.ylabel('Number of Movies')
plt.show()
```

 From the above visualizations
* You can see a genereal increase in movie release year after year showing an increase in demand
* Number of program releases differs by months. October and April are the months which have the highest number of program releases.
* From 2016 there has been a trend to atleast have new content out every month and this is expected to continue that way as they try to retain and add new movie lovers.
