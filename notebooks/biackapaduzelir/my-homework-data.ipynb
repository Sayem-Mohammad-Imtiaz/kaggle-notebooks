```python
df.chol.plot(kind='line', color='purple', label='chol', linewidth=1, alpha=1, grid=True, linestyle=':')
df.trestbps.plot(color='red', label= 'trestbps', linewidth=1, alpha=1, grid=True, linestyle='--')
plt.legend(loc= 'best')
plt.xlabel('x axis') #chol
plt.ylabel('y axis') #trestbps
plt.show()
```

mg / dl cinsinden serum kolesterol (Chol) ve Dinlenme kan basıncı (Trestbps) frekansları yukarıda gösterilmiştir. Chol değerlerimiz yaklaşık 100-300 aralığında değişim gösterir. Trestbps değerlerimiz yaklaşık 150-500+ aralığında değişim gösterir.

```python
df.plot(kind='scatter', x='thalach', y='restecg', alpha= 0.5, color='blue')
plt.xlabel('thalach')
plt.ylabel('restecg')
```
