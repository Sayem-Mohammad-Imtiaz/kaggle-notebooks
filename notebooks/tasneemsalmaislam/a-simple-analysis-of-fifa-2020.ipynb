A function to find of the index of desired columns index,

```python
def column_index(df, df_cols): 
    cols = df.columns.values
    ref = np.argsort(cols)
    return ref[np.searchsorted(cols,df_cols,sorter=ref)]
```

```python
column_index(df, ['age','club','contract_valid_until','dob','file','height_cm','international_reputation', #finding index of all the desired column
                  'joined','long_name','player_positions','nationality','overall','potential','short_name','weight_kg'])
```

***Question 1 : How many different nationality and clubs are there in FIFA 20 ?***

Answer : There are 162 different nationlities and 698 different club in FIFA 20. 

```python


count = 0

country = []

for i in range(0, len(new_df['nationality'])): #for loop to append all the unique countries/nationalites 
    
    if new_df['nationality'][i] not in country:  
        
        country.append(new_df['nationality'][i]) 
          
        count += 1
  
print(f'There are',count,'different nationalites in FIFA. ') 


print(f'The names of the different countries are: ')
print(country) #printing all the name of the countries that has a club/league
```

```python
count_club = 0 #number of distinct clubs in FIFA 2020 

league = []

for i in range(0, len(new_df['club'])): 
    
    if new_df['club'][i] not in league:  
        
        league.append(new_df['club'][i]) 
          
        count_club += 1
  
print(f'There are' ,count_club, 'different clubs in FIFA 20.') #number of unique clubs
```

***Question 5 : Which player(s) has the highest BMI and lowest BMI?***

Answer: Body Mass Index (BMI) is a person’s weight in kilograms divided by the square of height in meters. A high BMI can be an indicator of high body fatness.A. Akinfenwa from Wycombe Wanderers club has the highest BMI of 34.72 and J. Carrascal from River Plate has the lowest BMI. BMI under 18.5 is considered as underweight and over 29.9 is considered as overweight. 



```python
print('Players with the highest and lowest_BMI')

new_height = new_df['height_cm']/100

BMI = round(new_df['weight_kg']/(new_height**2),2)

new_df['BMI']= BMI

highest_BMI = new_df.sort_values(by = 'BMI', ascending = False)

player_h_bmi = highest_BMI[['short_name','nationality','club','BMI']]

H_BMI = player_h_bmi.head(1)

lowest_BMI = new_df.sort_values(by = 'BMI', ascending = True)

player_l_bmi = lowest_BMI[['short_name','nationality','club','BMI']]

L_BMI = player_l_bmi.head(1)

joined_BMI = pd.concat([H_BMI, L_BMI], axis=0)

print(joined_BMI)

cal_bmi = highest_BMI[['BMI']].agg(func=['min', 'mean', 'max'])

cal_bmi.plot(
    kind = 'line',
    ylabel = 'BMI', figsize =[12,8], title = 'Min, Max and AVG BMI'
    
)

print(cal_bmi)
```

***Question 7 : How many players plays at LW in FIFA 20?***

Answer : 378 players play in LW position. 

```python
import seaborn as sns
%matplotlib inline


All_positions = pd.DataFrame(new_df.player_positions.str.split(',',2).tolist(), 
                         columns = ['position1','position2','position3'])



player_position = All_positions.melt(var_name='position', value_name='LW_position')

player_position = player_position.dropna()

player_position['LW_position'] = player_position['LW_position'].str.replace(' ', '')

position = player_position.groupby(['LW_position']).count()

p_position = position.reset_index()



player_position.LW_position.value_counts().plot(kind = 'pie', subplots = True, figsize = [14,10],
title = 'Different position played by players')


plt.legend(title = 'position')
plt.show()

p_position[p_position['LW_position']=='LW']






```

The plot shows all the position that player play in FIFA 20. The number of players that plays in LW position are 981. The biggest portion in Pie chart is CB, most players play at CB positon. RWB is the lowest among all the postions.

***Question 9:  Is there any correlation between the overall score and potential with age and BMI in FIFA 20?***

Answer: Explained in below scatter matrix.

```python
corr_df = new_df[['overall','age','BMI','potential']]

pd.plotting.scatter_matrix(corr_df, alpha = .50, figsize = [12,8], )

plt.show

print('Covariance and Correlation of Age,BMI,Overall score and potential')

print(corr_df.corr())

print(corr_df.cov())
```

***Question 10: What is the highest number of players having the same nationality ?***

```python


fil_df = new_df[['nationality','short_name']]
player_n = fil_df.pivot_table(
    index = 'nationality',
    values = 'short_name',
    aggfunc = 'count'
)



nationality_p = player_n.sort_values(by = 'short_name', ascending = False)


print('Nationality that is common among players: ')
print(nationality_p.head(1))

high_df = fil_df.groupby('nationality')['short_name'].agg(func=['count'])

highest_df = high_df.reset_index()

highest_df.plot(
    kind = 'bar',
    x = 'nationality',
    y = 'count', figsize = [32,14], title = 'Countries with highest number of players')

plt.ylabel('Number of players')

plt.show()
```
