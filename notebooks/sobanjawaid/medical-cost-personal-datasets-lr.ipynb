```python
data = pd.read_csv("/kaggle/input/insurance/insurance.csv")
```

```python
data.head()
```

```python
y = data['charges'].to_frame()
```

```python
# label encoding the data 
from sklearn.preprocessing import LabelEncoder 
  
le = LabelEncoder() 
  
data['sex']= le.fit_transform(data['sex']) 
data['smoker']= le.fit_transform(data['smoker'])
data
```

```python
# label encoding the data 
from sklearn.preprocessing import LabelEncoder 
  
le = LabelEncoder() 
  
data['sex']= le.fit_transform(data['sex']) 
data['smoker']= le.fit_transform(data['smoker'])
data
```

```python
x = data
x=data.drop('region',axis=1)
x = x.drop('charges', axis=1)
type(x)
```

```python
# splitting data to training and testing dataset. 
from sklearn.model_selection import train_test_split 
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size =0.2, random_state = 0) 
print("xtrain shape : ", xtrain.shape) 
print("xtest shape : ", xtest.shape) 
print("ytrain shape : ", ytrain.shape) 
print("ytest shape : ", ytest.shape) 
```

```python
# Applying Linear Regression Model to the dataset and predicting the prices.

# Fitting Multi Linear regression model to training model 
from sklearn.linear_model import LinearRegression 
regressor = LinearRegression() 
regressor.fit(xtrain, ytrain) 

# predicting the test set results 
y_pred = regressor.predict(xtest)
```

```python
# Importing Libraries 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 

# Plotting Scatter graph to show the prediction results – ‘ytrue’ value vs ‘y_pred’ value

# Plotting Scatter graph to show the prediction 
# results - 'ytrue' value vs 'y_pred' value 
plt.scatter(ytest, y_pred, c = 'green') 
plt.xlabel("True charges") 
plt.ylabel("Predicted charges") 
plt.title("True value vs predicted value : Linear Regression") 
plt.show() 
```

```python
# Results of Linear Regression i.e. Mean Squred Error.

# Results of Linear Regression. 
from sklearn.metrics import mean_squared_error 
mse = mean_squared_error(ytest, y_pred) 
print("Mean Square Error : ", mse) 
```

```python
#############################################################################
###########.                    ACCURACY                           ##########
#############################################################################

accuracy = regressor.score(xtest,ytest)
accuracy
```
