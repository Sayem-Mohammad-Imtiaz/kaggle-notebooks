```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
data['Sex']=label.fit_transform(data['Sex'])
data['BP']=label.fit_transform(data['BP'])
data['Cholesterol']=label.fit_transform(data['Cholesterol'])
data['Drug']=label.fit_transform(data['Drug'])
```

```python
data.head()
```

```python
normalization_data=data.iloc[:,[0,4]]
```

```python
from sklearn.preprocessing import MinMaxScaler
minscaler=MinMaxScaler()
scaled=minscaler.fit_transform(normalization_data)
normalized_data=pd.DataFrame(scaled,columns=normalization_data.columns)
normalized_data
```

```python
normalized_dataset=pd.concat([normalized_data,data.iloc[:,[1,2,3,5]]],axis=1)
normalized_dataset
```

```python
input=normalized_dataset.iloc[:,:-1]
#input
output=normalized_dataset.iloc[:,-1]
output
```

```python
from sklearn.model_selection import train_test_split
input_train,input_test,output_train,output_test=train_test_split(input,output,test_size=0.2,random_state=250)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(random_state=90)
model.fit(input_train,output_train)
```

```python
predict=model.predict(input_test)
predict
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(predict,output_test)
accuracy
```
