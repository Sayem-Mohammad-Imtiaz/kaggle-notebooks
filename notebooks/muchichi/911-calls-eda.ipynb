**Breaking down a column (In this case title)** 

```python
df['title'].unique()[0]
```

```python
x = df['title'].iloc[0]
print(x)
print(x.split(':')[0])
```

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
df['Reason Detail'] = df['title'].apply(lambda title: title.split(':')[1])
df['Reason Detail'].head()
```

**Time information**
====================


```python
print(df['timeStamp'].iloc[0])
print(type(df['timeStamp'].iloc[0]))
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

```python
type(df['timeStamp'].iloc[0])
```

```python
time = df['timeStamp'].iloc[0]
time.dayofweek
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
```

```python
df.head()
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}  # Map attribute with other values}
```

```python
df['Day of Week'] = df['Day of Week'].map(dmap)
```

**Create a count plot of the Day of week with Reason as hue**

```python
sns.countplot(x='Day of Week',data=df,hue='Reason')
plt.legend(bbox_to_anchor=(1.05,1),loc=2, borderaxespad=0)
```

```python
sns.countplot(x='Month',data=df,hue='Reason')
plt.legend(bbox_to_anchor=(1.05,1),loc=2, borderaxespad=0)
```
