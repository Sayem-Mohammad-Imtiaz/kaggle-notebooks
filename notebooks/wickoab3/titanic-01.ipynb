```python
def fill_age(col):
    age = col[0]
    clas = col[1]
    
    if pd.isnull(age):
        if clas == 1:
            return titanic[titanic['Pclass'] == clas]['Age'].dropna().mean()
        elif clas == 2:
            return titanic[titanic['Pclass'] == clas]['Age'].dropna().mean()
        else:
            return titanic[titanic['Pclass'] == clas]['Age'].dropna().mean()
    else:
        return age
```

```python
titanic[['Age']] = titanic[['Age', 'Pclass']].apply(fill_age, axis = 1)
```

```python
titanic.head()
```

```python
sns.distplot(titanic['Age'])
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(titanic.isnull(), cbar =False, yticklabels = False, cmap = 'viridis')
```

```python
titanic.dropna(inplace = True)
```

```python
plt.figure(figsize = (12,8))
sns.heatmap(titanic.isnull(), cbar = False, yticklabels = False, cmap = 'viridis')
```

```python
titanic.info()
```

```python
sex = pd.get_dummies(titanic['Sex'], drop_first = True)
```

```python
embark = pd.get_dummies(titanic['Embarked'], drop_first = True)
```

```python
titanic.drop(['Sex', 'Embarked', 'Ticket'], axis = 1, inplace = True)
```

```python
train = pd.concat([titanic, sex, embark], axis = 1)
```

```python
train.head()
```

```python
def get_gen(name):
    if 'Mr.' in name :
        return 'Mr.'
    elif 'Mrs.' in name:
        return 'Mrs.'
    elif 'Miss.' in name:
        return 'Miss.'
    else:
        return 'not signed'
```

```python
train['gen'] = train['Name'].apply(get_gen)
```

```python
train.drop(['Name'], axis = 1, inplace = True)
```

```python
train.head()
```

```python
gen = pd.get_dummies(train['gen'], drop_first = True)
```

```python
train = pd.concat([train, gen], axis = 1)
```

```python
train.head()
```

```python
train.drop(['gen'], axis = 1, inplace = True)
```

```python
train.head()
```

# Time to Train

```python
from sklearn.model_selection import train_test_split
```

```python
x = train.drop('Survived', axis = 1)
y = train['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(x_train, y_train)
```

```python
prediction = logmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, prediction))
```
