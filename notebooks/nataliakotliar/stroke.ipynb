```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.figure_factory as ff
from plotly import tools
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.express as px
import seaborn as sns
#from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

plt.style.use('seaborn-whitegrid')
import warnings
warnings.filterwarnings('ignore')
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score
```

```python
dataset = pd.read_csv('/kaggle/input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')
dataset.head()
```

```python
dataset.describe(include = 'all')
```

```python
print(len(dataset))
dataset.info(memory_usage="deep")
```

```python
g1=0; g2=0
dataset['is_Stroke'] = ' '
# підрахунок у кого інсульт
for i in range(len(dataset)):
    if dataset['stroke'][i] == 1:
        dataset['is_Stroke'][i] = 'Yes'
        g1=g1+1
    else:
        dataset['is_Stroke'][i] = 'No'
        g2=g2+1
print(g1, g2)
```

```python
plt.figure(figsize = (7,5))
sns.countplot(x = dataset['gender'])
plt.title('Стать', fontsize = 20)
plt.xlabel('Стать')
plt.ylabel('Підрахунок')
plt.show()
```

```python
plt.figure(figsize = (7, 5))
sns.countplot(x = dataset['is_Stroke'])
plt.title('Інсульт', fontsize = 20)
plt.xlabel('Інсульт');plt.ylabel(' ')
plt.show()
```

```python
plt.figure(figsize = (7, 5))
sns.countplot(x = dataset['heart_disease'])
plt.title('Хвороба серця', fontsize = 20)
plt.xlabel('Хвороба серця');plt.ylabel(' ')
plt.show()
```

```python
plt.figure(figsize = (7, 5))
sns.countplot(x = dataset['ever_married'])
plt.title('Сімейний стан', fontsize = 20)
plt.xlabel('Одружений/а');plt.ylabel(' ')
plt.show()
```

```python
plt.figure(figsize = (7, 5))
sns.countplot(x = dataset['work_type'])
plt.title('Тип роботи', fontsize = 20)
plt.xlabel('Робота');plt.ylabel(' ')
plt.show()
```

```python
plt.figure(figsize = (7, 5))
sns.countplot(x = dataset['Residence_type'])
plt.title('Зона проживання', fontsize = 20)
plt.xlabel('міська/сільська');plt.ylabel(' ')
plt.show()
```

```python
plt.figure(figsize = (7, 5))
sns.countplot(x = dataset['smoking_status'])
plt.title('Паління', fontsize = 20)
plt.xlabel('Паління');plt.ylabel(' ')
plt.show()
```

```python
plt.figure(figsize = (10, 7))
sns.barplot(x = dataset['smoking_status'], y = dataset['heart_disease'], hue = dataset['work_type'])
plt.legend(loc = 'upper right')
plt.title('Палії vs хвороб серця на основі типу роботи', fontsize = 20)
plt.xlabel('Палії')
plt.ylabel('Хвороби серця')
plt.show()
```
