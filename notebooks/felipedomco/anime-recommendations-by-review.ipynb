Usando Correlação para sugerir animes a partir de sua Avaliação dos usuarios.

# 1-) Importando bibliotecas necessarias.

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
%matplotlib inline
```

# 2-) Carregando a base de dados de animes

```python
df = pd.read_csv('../input/anime-recommendations-database/anime.csv')
df.head()
```

# 3-) Carregando a base de Avaliações de Usuarios

```python
dfr = pd.read_csv('../input/anime-recommendations-database/rating.csv')
dfr.head()
```

# 4-) Analise dos dados fornecidos.

Como o objetivo é usar as avalições dos usuarios, foram retirados as avaliações com "-1", que são usuarios que não deram nota. Diferente de avaliação "0", que é quando o usuario deu a nota minima para o anime.

```python
dfr = dfr[dfr['rating'] >= 0]
dfr.head()
```

Realizando o merge entre os 2 dataframes, Animes e Avaliações a partir do campo anime_id.

```python
df = pd.merge(df,dfr,on='anime_id')
df.head()
```

Visualização dos animes mais avaliados.

```python
df.groupby('name')['rating_y'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('name')['rating_y'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('name')['rating_y'].mean())
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('name')['rating_y'].count())
ratings.head()
```

Grafico agrupado com as maiores quantidades de avaliações.

```python
plt.figure(figsize=(10,4))
ratings['num of ratings'].hist(bins=70)
```

Grafico com agrupamento das avaliações.

```python
plt.figure(figsize=(10,4))
ratings['rating_y'].hist(bins=70)
```

```python
sns.jointplot(x='rating_y',y='num of ratings',data=ratings,alpha=0.5)
```

Criando uma tabela onde detalha as avaliações por usuario.

```python
animes = df.pivot_table(index='user_id',columns='name',values='rating_y')
animes.head()
```

Ranking dos animes mais avaliados.

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

Indicação de anime para Dragon Ball Z

```python
anime_user_ratings = animes['Dragon Ball Z']
similar_to_anime = animes.corrwith(anime_user_ratings)
corr_anime = pd.DataFrame(similar_to_anime,columns=['Correlation'])
corr_anime.dropna(inplace=True)
corr_anime = corr_anime.join(ratings['num of ratings'])
corr_anime[corr_anime['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```

Indicação de anime para Shingeki no Kyojin

```python
anime_user_ratings = animes['Shingeki no Kyojin']
similar_to_anime = animes.corrwith(anime_user_ratings)
corr_anime = pd.DataFrame(similar_to_anime,columns=['Correlation'])
corr_anime.dropna(inplace=True)
corr_anime = corr_anime.join(ratings['num of ratings'])
corr_anime[corr_anime['num of ratings']>100].sort_values('Correlation',ascending=False).head()
```
