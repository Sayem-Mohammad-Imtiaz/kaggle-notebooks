```python
#!/usr/bin/env python
# coding: utf-8

# In[50]:
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
get_ipython().run_line_magic('matplotlib', 'inline')
# In[61]:
stud_df = pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
# In[52]:
log_gender = pd.get_dummies(stud_df['gender'], drop_first=True)
log_race = pd.get_dummies(stud_df['race/ethnicity'], drop_first=True)
log_parental = pd.get_dummies(stud_df['parental level of education'], drop_first=True)
log_lunch = pd.get_dummies(stud_df['lunch'], drop_first=True)
log_prep = pd.get_dummies(stud_df['test preparation course'], drop_first=True)
# In[53]:
stud_df = pd.concat([stud_df, log_gender, log_race, log_parental, log_lunch, log_prep], axis=1)
# In[54]:
stud_df.drop(['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course'], axis=1, inplace=True)
# In[56]:
X = stud_df.drop('male', axis=1)
y = stud_df['male']
# In[57]:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
# In[58]:
logmodel = LogisticRegression(max_iter=10000)
logmodel.fit(X_train,y_train)
# In[59]:
pred = logmodel.predict(X_test)
# In[60]:
print(classification_report(y_test, pred))
```
