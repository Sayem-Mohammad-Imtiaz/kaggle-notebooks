```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import folium
```

```python
df = pd.read_csv("../input/volcanic-eruptions/database.csv")
```

```python
df["Region"].value_counts().plot(kind = "bar", rot = 45, figsize = (15,5), title = "Volcanos in different regions")
```

```python
plt.hist(df["Elevation (Meters)"])
plt.title("Distribution of elevation")
plt.ylabel("Frequency")
plt.xlabel("Elevation")
```

```python
df["Dominant Rock Type"].value_counts().plot(kind = "bar", rot = 45, figsize = (15,5), title = "Frequency of rock types")
```

```python
dates = df["Last Known Eruption"]
```

```python
new_dates = []
for i in range(len(dates)):
    new_date = dates[i].split(" ")
    if len(new_date) == 1:
        continue
    elif "BCE" in new_date:
        new_dates.append("-" + new_date[0])
    elif "CE" in new_date:
        new_dates.append(new_date[0])
```

```python
new_dates[0:5]
```

```python
plt.hist(list(map(int, new_dates)))
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.title("Histogram of eruption dates")
```

```python
df["Activity Evidence"].value_counts().plot(kind = "bar", rot = 45, title = "Frequencies of evidence of eruptions")
```

## Location of volcanos

```python
m = folium.Map(location=[0, 0], zoom_start=2)
labels = pd.factorize(df["Region"])[0]
loc = 'Volcanos across the world'
title_html = '''
             <h3 align="center" style="font-size:16px"><b>{}</b></h3>
             '''.format(loc)   
for i in range(0,len(df)):
    folium.CircleMarker(
      location=[df.iloc[i]['Latitude'], df.iloc[i]['Longitude']],
      radius=float(abs(df.iloc[i]["Elevation (Meters)"]))/2000,
      popup = df.iloc[i]["Name"],
      color = float(labels[i]),
      fill=True,
      fill_color= "crimson"
   ).add_to(m)
m.get_root().html.add_child(folium.Element(title_html))
folium.LayerControl().add_to(m)
m
```
