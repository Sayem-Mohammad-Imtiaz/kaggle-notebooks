```python
from sklearn.linear_model import LinearRegression
```

```python
def country(x):
    country = x;
    return country
```

```python
ct = base['country'].map(country)
```

```python
ctu = [];
def unico(ct):
    for i in ct:
        if i not in ctu:
            ctu.append(i);
```

```python
unico(ct)
```

```python
def geracao(x):
    if x == '5-14 years':
        return 1
    elif x == '15-24 years':
        return 2
    elif x == '25-34 years':
        return 3
    elif x == '35-54 years':
        return 4
    elif x == '55-74 years':
        return 5
    elif x == '75+ years':
        return 6
    else:
        return 0
```

```python
base['idade'] = base['age'].map(geracao)
```

```python
def countries(i):
    country = base[base['country'] == i]
    y = country['suicides_no']
    country = country.drop(columns=['year','suicides_no','country', 'sex', 'age', 'country-year',' gdp_for_year ($) ', 'generation'])
    X =  country
    X.fillna(0.0, inplace = True)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
    lm = LinearRegression()
    lm.fit(X_train,y_train)
    coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
    print(i)
    print(coeff_df)
    print('\n')
```

```python
for i in ctu:
    countries(i);
```
