```python
dataFrame = pd.read_csv("../input/kc_house_data.csv")
dataFrame.head()
```

```python
dataFrame.info()
```

```python
dataFrame.describe()
```

```python
dataFrame = dataFrame.drop(["id", "date"], axis= 1)
```

```python
f,ax = plt.subplots(figsize=(16, 16))
sns.heatmap(dataFrame.corr(), annot=True, linewidths=0.5,linecolor="red", fmt= '.1f',ax=ax)
plt.show()
```

```python
data_for_plotting = dataFrame[["sqft_living", "bedrooms", "sqft_lot", "sqft_above", "grade", "price"]]
plt.figure(figsize=(12,8))
sns.pairplot(data_for_plotting, diag_kind= "hist", height= 3)
plt.show()
```

```python
#dimension for regression : sqft_living, grade, sqft_above, sqft_living15
y = dataFrame.price.values.reshape(-1,1)
x_data = dataFrame.drop(["price"], axis = 1)

#normalization
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data)).values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test,y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state = 42)
```

# Lineer Regression

```python
from sklearn.linear_model import LinearRegression

linear_reg = LinearRegression()
linear_reg.fit(x_train["sqft_living"].values.reshape(-1,1),y_train)
from sklearn.metrics import r2_score
y_head = linear_reg.predict(x_test["sqft_living"].values.reshape(-1,1))
print("R2 score:", r2_score(y_test, y_head))
```

```python
#plotting
prediction_line = (np.arange(np.min(x_train["sqft_living"]),np.max(x_train["sqft_living"]),
                   (np.max(x_train["sqft_living"]) -np.min(x_train["sqft_living"]))/100 )).reshape(-1,1)
y_head = linear_reg.predict(prediction_line)

plt.figure(figsize=(8,6))
plt.plot(prediction_line, y_head, color = "blue", label = "Prediction")
plt.scatter(x_train["sqft_living"],y_train, color="red", alpha = 0.2, label= "Train data")
plt.legend()
plt.xlabel("Sqft")
plt.ylabel("Price")
plt.show()
```

# Multiple Linear Regression

```python
#i select sqft_living, bedrooms, bathrooms, grade, sqft_above 
x_mlr_train = x_train[["sqft_living", "bedrooms", "bathrooms", "grade", "sqft_above"]]
x_mlr_test = x_test[["sqft_living", "bedrooms", "bathrooms", "grade", "sqft_above"]]
```

```python
multiple_lineer_regression = LinearRegression()
multiple_lineer_regression.fit(x_mlr_train, y_train)

print("Score : ", multiple_lineer_regression.score(x_mlr_test, y_test))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor

decision_tree = DecisionTreeRegressor()
decision_tree.fit(x_train, y_train)
y_head = decision_tree.predict(x_test)
print("R2 score", r2_score(y_head, y_test))
```
