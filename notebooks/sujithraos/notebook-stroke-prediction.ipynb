```python
df = pd.read_csv('../input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')
```

```python
print(df.info())
print(df.shape)
print(df.head())
print(df.describe())
```

```python
attr = ['gender','ever_married','work_type','Residence_type','smoking_status']
for i in attr:
    print (i)
    print(df[i].unique())
```

# Mapping to numbers

```python
df_cleaned=df
df_cleaned['gender_id'] = df_cleaned['gender'].map({'Male':0,'Female':1,'Other':2})
df_cleaned['ever_married_id'] = df_cleaned['ever_married'].map({'Yes':1,'No':0})
df_cleaned['work_type_id'] = df_cleaned['work_type'].map({'Private':0,'Self-employed':1,'Govt_job':2,'children':3,'Never_worked':4})
df_cleaned['Residence_type_id'] = df_cleaned['Residence_type'].map({'Urban':1,'Rural':0,})
df_cleaned['smoking_status_id'] = df_cleaned['smoking_status'].map({'formerly smoked':0,'never smoked':1,'smokes':2,'Unknown':3})                                                  
```

```python
df_cleaned.info()
```

```python
df_cleaned = df_cleaned.drop(columns=['gender','ever_married','work_type','Residence_type','smoking_status'])
```

```python
df_cleaned.info()
```

```python
cm = df_cleaned.corr()
print(cm['stroke'])
```

```python
df_cleaned = df_cleaned.drop(columns=['id'])
```

```python
mean_bmi = df_cleaned['bmi'].mean()
print(mean_bmi)
```

```python
df_cleaned['bmi'] = df_cleaned['bmi'].fillna(mean_bmi)
df_cleaned.info()
```

```python
df_cleaned_final = df_cleaned
```

```python
y=df_cleaned_final['stroke']
df_cleaned_final.drop(columns='stroke')
df_cleaned_scaled=df_cleaned_final
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = MinMaxScaler()
df_cleaned[['age','hypertension','heart_disease','avg_glucose_level','bmi','gender_id','ever_married_id','work_type_id','Residence_type_id','smoking_status_id']] = scaler.fit_transform(df_cleaned[['age','hypertension','heart_disease','avg_glucose_level','bmi','gender_id','ever_married_id','work_type_id','Residence_type_id','smoking_status_id']])
```

```python
X=df_cleaned
X.info()
y=df_cleaned['stroke']
X=X.drop(columns=['stroke'])
X.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier()
tree_clf.fit(X_train, y_train)
y_lbl_predicted = tree_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_lbl_predicted)
print(acc)
```

```python
from sklearn.ensemble import RandomForestClassifier
#rnd_clf = RandomForestClassifier(n_estimators=500, max_leaf_nodes=16, n_jobs=-1)
rnd_clf = RandomForestClassifier()
rnd_clf.fit(X_train, y_train)
y_pred_rf = rnd_clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred_rf)
print(acc)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
#bag_clf = BaggingClassifier(DecisionTreeClassifier(), n_estimators=500,max_samples=100, bootstrap=True, n_jobs=-1)
bag_clf = BaggingClassifier(DecisionTreeClassifier(), bootstrap=True, n_jobs=-1)
bag_clf.fit(X_train, y_train)
y_pred_bgcf = bag_clf.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_pred_bgcf)
print(acc)
```
