## Plotting the sightings by state

```python
def plot(x,y):
    plt.subplots(figsize=(30,10))
    sns.barplot(x=x,y=y,palette="GnBu_d")
```

### Changing to the date format and then plotting

```python
ufo_dates = df.datetime.str.replace('24:00', '00:00')
ufo_dates = pd.to_datetime(ufo_dates, format='%m/%d/%Y %H:%M')

df['datetime'] = df.datetime.str.replace('24:00', '00:00')

df['datetime'] = pd.to_datetime(df['datetime'], format='%m/%d/%Y %H:%M')
```

```python
year=ufo_dates.dt.year
```

```python
year_counts=year.value_counts()
year_labels=year_counts.index
year_freq=year_counts.values
```

```python
plot(year_labels,year_freq)
```

### Seems like only some last values are siginificant

```python
plot(year_labels[:50],year_freq[:50])
```
