```python
# 1. Data Preprocessing

#Importing Libraries
import numpy as np # linear algebra
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression


#Importing dataset
pd = pd.read_csv('/kaggle/input/housesalesprediction/kc_house_data.csv')

#Checking for datatypes
pd.dtypes
```

```python
#Checking the headers
pd.head()
```

```python
#Defining X and Y
X = pd['sqft_above']
Y = pd['sqft_living']

#Transforming X from int64 to int 32
X = X.astype(np.int32)
print( 'X is dtype', X.dtypes)
```

```python
#Transforming Y from int64 to int32
Y = Y.astype(np.int32)
print( 'Y is dtype', Y.dtypes)
```

```python
#Find the type of variable X
type(X)
```

```python
#Transforming X from Pandas Series to Numpy Array
X = X.values
type(X)
```

```python
#Transforming X 1D Numpy Array to 2D Numpy Array since that's expected by the LinearRegression
X = X.reshape(-1, 1) # reshaping like this because the data has a single feature
type(X)
#Splitting the data into Training set and Test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=100)
```

```python
# 2. Train and Test Data

#Creating the instance of the Linear Regression
regr = linear_model.LinearRegression()
```

```python
# 3. Training the model
regr.fit(X_train, Y_train)
```

```python
# 4. Testing the model
Y_pred = regr.predict(X_test)
```

```python
# 5. Evaluation

# The coefficients
print('Coefficients: \n', regr.coef_)
#The coefficion of prediction

# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(Y_test, Y_pred))
#The mean square error is very hign and far away from zero (perfect accuracy).

# The coefficient of determination
print('Coefficient of determination (R squared): %.2f'
      % r2_score(Y_test, Y_pred))
#Based on the coefficient of determination, only 77% of <sqft_living> can be predicted by <sqft_above>, which means it's an incorrect prediction model 
#since it's below 90-95%.

# Creating plot
plt.scatter(X_test, Y_test,  color='black')
plt.plot(X_test, Y_pred, color='red', linewidth=2)

plt.xticks(())
plt.yticks(())
plt.show()
```
