# Reading the CSV File üìÅ

# Preprocessing ‚öôÔ∏è

```python
train_df = df.select_dtypes(exclude=['int','float'])
```

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
for i in train_df:
    train_df[i] = ord_enc.fit_transform(train_df[[i]])
```

```python
for i in train_df:
    df[i] = train_df[i]
```

```python
col_train = list(df.columns)
prediction_set = df['class']
df.drop('class',axis = 1, inplace = True)
training_set = df
FEATURES = list(df.columns)
```

# **Train Test division** ‚ûó

```python
x_train, x_test, y_train, y_test = train_test_split(training_set , prediction_set, test_size=0.15, random_state=33)
```

```python
print(x_train.shape)
```

# MLP model üëª

```python
model = Sequential()
model.add(Dense(64, input_dim=21, activation='relu'))
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(256, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.3))
model.add(Dense(1, activation='sigmoid'))
model.summary()
model.compile(optimizer='adam',
                 loss='binary_crossentropy', metrics=['acc'])
```

```python
model.fit(x_train, y_train, epochs=30, validation_data=(x_test, y_test), batch_size=28, verbose=1)
```

```python
y = model.predict(x_test)
```

```python
final = pd.DataFrame(y_test)
final['prdicted'] = y
final['actual'] = final['class']
final = final.drop('class', axis = 1)
```
