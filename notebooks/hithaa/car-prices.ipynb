```python
data=pd.get_dummies(data,drop_first=True)
data.head()
```

```python
data.head()
```

```python
x=data.drop(['Selling_Price','Present_Price'],axis=1)
x.head()
```

```python
y=data[['Selling_Price']]
y.head()
```

```python
y.Selling_Price=LabelEncoder().fit_transform(y.Selling_Price)
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train
```

```python
x_test.shape
```

```python
x_train.head()
```

```python
trans_nom=make_column_transformer((OneHotEncoder(sparse=False),[0,3,5]),(OrdinalEncoder(),[2,4]),remainder="passthrough")
```

```python
trans_nom.fit_transform(x_train)
```

```python
model=LogisticRegression(solver='liblinear')
```

```python
set_config(display='diagram')
pipe=make_pipeline(trans_nom,model)
pipe
```

```python
pipe.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```
