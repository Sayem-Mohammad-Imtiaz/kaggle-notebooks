```python
#LINE PLOT
#color = color, label =label, linewidth = widht of line, alpha = opacity grid = grid, linestyle = style
#alpha 0.5 yani saydamlığı, grid o arka plandaki kareler, linestyle : tire olabilir 
#herbir x ekseni pokemonlar
#her bir y ekseni hız ve defans

data.Speed.plot(kind = 'line', color = 'g', label = 'Speed', linewidth=1, alpha = 0.5, grid= True, linestyle = ':')
data.Defense.plot(color = 'r', label = 'Defense', linewidth=1, alpha = 0.5, grid = True, linestyle = '-.')
plt.legend(loc='upper right')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Line plot')
plt.show()
```

```python
#SCATTER PLOT
#amaç 2 feature 2 variable arasındaki ilişkiyi correlationu bulmak
#correlatşon neydi d.o mı t.o mı 
#benim burda öğrenmek istediğim ne attack ve defense arasındaki ilişki ne
#ben atağı x eksenşne defense'i y eksenime koyuyorum.

data.plot(kind='scatter', x ='Attack',y= 'Defense', alpha = 0.5, color ='red')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Attack Defense Scatter Plot')
plt.show()
#Attack büyük harfle datanın attack'ı DİKKAT
#alpha 0.5 saydam yaptı
```

```python
#HISTOGRAM
# bins = number of bar in figure
data.Speed.plot(kind= 'hist', bins=50, figsize=(15,15))
plt.show()
#50 hızına sahip pokemonların sayısı örneğin (frekansı)
#bin barların sayısı / çubuklar
#figure size : frame'in boyutu
```

```python
#create dictionary and look its keys and values
dictionary = {'spain' : 'madrid', 'usa': 'vegas'}
print(dictionary.keys())
print(dictionary.values())
```

```python
#keys have to be immutable objects like string, boolean,float, integer or tubles
#List is not immutable
#keys are unique

dictionary['spain'] = "barcelona"
#update existing entry
print(dictionary)

dictionary['france'] = "paris"
#add new entry
print(dictionary)

del dictionary['spain']
#remove entry with key 'spain'
print(dictionary)

print('france' in dictionary)
#check include or not
dictionary.clear() #remove all entries
print(dictionary)
```

```python
data = pd.read_csv('../input/pokemon-challenge')
```

```python
#PANDAS 2 çeşit data typedan oluşuyor.
#1. Series
#2. DataFrame
#serilerle dataframe arasındaki fark: seriler numpy'dan hatırlarsınız bir vektör şeklindqe uzanan
#tek boyutlu yapılardı ve bunların metodları dataframe'in metodlarından biraz daha farklı.
#seri oluşturmak için datanın defensini alıyorum, bir seri oluşuyo, ama datanın bir liste içinde defensini alırsam dataframe oluşuyo.

series = data['Defense']
print(type(series))
data_frame = data[['Defense']]
print(type(data_frame))
```

```python
# 1 - Filtering Pandas data frame
x = data['Defense']>200
data[x]
```

```python
# 2- Filtering Pandas with Logical_and
#iki tarafta doğru olmalı
data[np.logical_and(data['Defense']>200, data['Attack']>100)]
#logical_and numpy kütüphanesinde var.
#ikisinin de doğru olmasını istediğim için logical_and kullandım.
#data'ma soktuğum zaman false'lari ignore ediyor ve yazdırmıyor.
```

```python
#This is also same with previous code line. Therefore we can also use '&' for filtering.
data[(data['Defense']>200) & (data['Attack']>100)]
#VE
```

WHILE AND FOR LOOPS

```python
i=0
while i != 5:
    print('i is :',i)
    i +=1
print(i, 'is equal to 5')
```

```python
#lis'in içinde i dolansın.
lis = [1,2,3,4,5]
for i in lis:
    print('i is:',i)
print('')

#Enumerate: listenin içindeki değerlerin indexlerine de ulaşmak için

for index, value in enumerate(lis):
    print(index, " : ",value)
print('')

#For dictionaries
#dictionary'nin içine erişebilmek için for kullanabilirim.
dictionary= {'spain':'madrid', 'france': 'paris'}
for key, value in dictionary.items():
    print(key, ":", value)
print('')

#pandasa erişim sağlamak için
#iterrows: atağımı buluyo ayrıca indexini de buluyor.

for index,value in data[['Attack']][0:1].iterrows():
    print(index ,":",value)
```
