```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
call  = pd.read_csv('/kaggle/input/montcoalert/911.csv')
```

**Number of unique title codes**

```python
call['title'].nunique()
```

**Created a new column 'Reason' based off the title of the calls**

```python
call['Reason'] = call['title'].apply(lambda x:x.split(':')[0])
```

**The most common Reason for a 911 call based off the new column**

```python
call['Reason'].value_counts()
```

```python
sns.countplot(x='Reason',data=call, palette = 'viridis')
```

**Checking the data type of timeStamp**

```python
type(call['timeStamp'].iloc[0])
```

**Converting timeStamp from str to datetime**

```python
call['timeStamp'] = pd.to_datetime(call['timeStamp'])
```

```python
type(call['timeStamp'].iloc[0])
```

```python
call['timeStamp']
```

**You can now grab specific attributes from a Datetime object by calling them.**

```python
time = call['timeStamp'].iloc[0]
time.hour
# time.dayofweek
# time.month
# time.date()
```

**Created 3 new columns for Hour, Month and Day of Week**

```python
call['Hour'] = call['timeStamp'].apply(lambda time: time.hour)
call['Month'] = call['timeStamp'].apply(lambda time:time.month)
call['Day of Week'] = call['timeStamp'].apply(lambda time:time.dayofweek)
```

```python
call.sample(5)
```

**Used the .map to map the actual string names to the day of the week**

```python
dmap={0:'Mon', 1:'Tue', 2:'Wed', 3:'Thu', 4:'Fri', 5:'Sat', 6:'Sun'}
```

```python
call['Day of Week'] = call['Day of Week'].map(dmap)
call.head()
```

**Countplot of Day of the Week based off the Reason column**

```python
sns.countplot(x='Day of Week',data=call, hue='Reason',palette='viridis')
plt.legend(bbox_to_anchor=(1.05, 1))
```

**Countplot of Month based off the Reason column**

```python
sns.countplot(x='Month',data=call, hue='Reason',palette='viridis')
plt.legend(bbox_to_anchor=(1, 1))
```

```python
byMonth = call.groupby('Month').count()
byMonth
```

**Simple plot indicating the count of calls per month. It clearly shows a big drop in calls in August.**

```python
sns.pointplot(x='Month', y = 'lat', data=byMonth.reset_index(), markers='o')
```

```python
t = call['timeStamp'].iloc[0]
t.date()
```

**Created a new column 'Date' from the TimeStamp column.**

```python
call['Date'] = call['timeStamp'].apply(lambda t: t.date())
call['Date']
```

```python
call.sample(5)
```

```python
call.groupby('Date').count().head()
```

**Simple countplot of Total 911 calls**

```python
call.groupby('Date').count()['lat'].plot()
```

**Plot representing 'EMS' Reason for the 911 call.**

```python
call[call['Reason']=='EMS'].groupby(by='Date').count()['lat'].plot()
plt.title('EMS')
```

**Plot representing 'Fire' Reason for the 911 call.**

```python
call[call['Reason']=='Fire'].groupby(by='Date').count()['lat'].plot()
plt.title('Fire')
```

**Plot representing 'Traffic' Reason for the 911 call.**

```python
call[call['Reason']=='Traffic'].groupby(by='Date').count()['lat'].plot()
plt.title('Traffic')
```

**Created a new dataframe to clearly show the number of calls taken per hour everyday.**

```python
dayHour = call.groupby(by=['Day of Week', 'Hour']).count()['Reason'].unstack()
dayHour
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour)
```

**Created a new dataframe to clearly show the number of calls taken per day of every month.**

```python
dayMonth = call.groupby(by=['Day of Week', 'Month']).count()['Reason'].unstack()
dayMonth
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth)
```
