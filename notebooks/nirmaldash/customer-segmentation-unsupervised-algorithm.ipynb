```python
X=df.iloc[:,[3,4]].values
```

Finding the optimum number of cluster using Dendogram

```python
plt.figure(figsize=(12,6))
from scipy.cluster import hierarchy as sch
dendro=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title("Dendogram")
plt.ylabel("Euclidean Distance")
plt.xlabel("Customer")
plt.show()
```

From the above Dendogram image we can select line where we can get the cluster that cosidering almost all data points. The distance which is cover mode is 4, so we can take 5 clusters.

```python
#Training Hierarchical Clustering

from sklearn.cluster import AgglomerativeClustering
AGC = AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_pred=AGC.fit_predict(X)
```

```python
plt.figure(figsize=(12,6))
#Visualize the Cluster
plt.scatter(X[y_pred == 0,0],X[y_pred == 0,1], s = 100, c = 'blue', label = 'Cluster 1')
plt.scatter(X[y_pred == 1,0],X[y_pred == 1,1], s = 100, c = 'green', label = 'Cluster 2')
plt.scatter(X[y_pred == 2,0],X[y_pred == 2,1], s = 100, c = 'red', label = 'Cluster 3')
plt.scatter(X[y_pred == 3,0],X[y_pred == 3,1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[y_pred == 4,0],X[y_pred == 4,1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.title("Cluster of Customers")
plt.xlabel("Annual Income of Customer")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```

--Finding the optimal number of clusters using the elbow method

```python
#Import All Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/kaggle/input/mall-customers/Mall_Customers.csv')
```

```python
print(df.shape)
print("\n")
print(df.info())
```

```python
#Extracting Independent Variable
X=df.iloc[:,[3,4]].values
```

```python
#Step-2: Finding the optimal number of clusters using the elbow method
from sklearn.cluster import KMeans
wcss = [] #Initialise the list for value of wcss

#Take K value 10 and iterate 1 to 10 on K-Means algorithm and see how this is working

for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title("The Elbow method graph")
plt.xlabel("The Number of Cluster(K)")
plt.ylabel("Within Cluster Sum Squared")
plt.show()
```

From the above graph it is clear that the elbow point is at 5, so number of cluster is required is 5

```python
[np.round(i,2) for i in wcss]
```

```python
#Apply K-Means algorithm on training dataset.

kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_pred=kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(12,6))
plt.scatter(X[y_pred == 0,0],X[y_pred == 0,1],s=100,c='blue',label='Cluster-1')
plt.scatter(X[y_pred == 1,0],X[y_pred == 1,1],s=100,c='green',label='Cluster-2')
plt.scatter(X[y_pred == 2,0],X[y_pred == 2,1],s=100,c='red',label='Cluster-3')
plt.scatter(X[y_pred == 3,0],X[y_pred == 3,1],s=100,c='cyan',label='Cluster-4')
plt.scatter(X[y_pred == 4,0],X[y_pred == 4,1],s=100,c='magenta',label='Cluster-5')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,c='yellow',label='Centroid')
plt.title("K-Means Clustering")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```

### What we observe from above patterns

**Cluster-1 : Customer in this category having average annual income and average spending in shopping mall, these are medium class customer**

**Cluster-2 : Customer in this category having high annual income but less spending in shopping mall. So we can category them as "Careful"**

**Cluster-3 : Customer in this category having less annual income and less spending, so we can categorized them as "Sensible"**

**Cluster-4 : Customer in thsi category less annual income but high spending in shopping mall, we can categorized them as "Careless"**

**Cluster-5 : Customer in thsi category having high annual income and high spending in shopping mall, this customer has categorized as "Rich", so shopping mall can target these customer to get profit**
