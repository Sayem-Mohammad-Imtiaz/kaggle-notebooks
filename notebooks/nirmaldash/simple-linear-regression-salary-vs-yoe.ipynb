```python
df=pd.read_csv('/kaggle/input/salary-data-simple-linear-regression/Salary_Data.csv')
```

```python
df.info()
```

```python
#Independent variable : years of exprience
#Dependent Variable : Salary
#Visualize how my dependent and independent variables are scattered
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(df['YearsExperience'],df['Salary'])
plt.plot(df['YearsExperience'],df['Salary'],color='red')
plt.title("Salary vs Experience (Training Dataset)") 
plt.xlabel("Years of Experience")  
plt.ylabel("Salary(In Rupees)")  
plt.show() 
```

```python
#Check the outlier
plt.boxplot(df['YearsExperience'],patch_artist=True)
plt.show()
```

```python
#Identify dependent and independent variable
#X=df.iloc[:,:-1].values
#y=df.iloc[:,1].values
X=df[['YearsExperience']]
y=df['Salary']
print(X.shape)
print(y.shape)
```

```python
#Split X and y into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=41)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Linear Regression model creation and fit training data and validate with test data
from sklearn.linear_model import LinearRegression
model_lr=LinearRegression()
model_lr.fit(X_train,y_train)
print(model_lr.coef_)
print(model_lr.intercept_)
```

```python
#Validate the model with test and train
y_pred=model_lr.predict(X_test)
x_pred=model_lr.predict(X_train)
```

```python
#visualizing the Training set results:
plt.scatter(X_train,y_train,color="green")
plt.plot(X_train,x_pred,color="red")
plt.title("Salary vs Experience (Training Dataset)")
plt.xlabel("Years of Experience")  
plt.ylabel("Salary(In Rupees)")  
plt.show()  
```

```python
#visualizing the Test set results:
plt.scatter(X_test,y_test,color="blue")
plt.plot(X_test,y_pred,color="red")
plt.title("Salary vs Experience (Testing Dataset)")
plt.xlabel("Years of Experience")  
plt.ylabel("Salary(In Rupees)")  
plt.show()  
```

```python
#Validate the model with Unseen or New data
YOE=[12,10,32,14,15,17,18,19,6,4]
YOE=np.reshape(np.array(YOE),newshape=(len(YOE),1))
prediction=np.round(model_lr.predict(YOE),2)
NewDataFrame=pd.DataFrame(YOE,columns=['Years Of Exprience'])
NewDataFrame['Salary']=prediction
print(NewDataFrame)
```

```python
#Visualize New data with theire predicted Salary
plt.scatter(YOE,prediction,color="green")
plt.plot(YOE,prediction,color="red")
plt.title("YOE Vs Salary Prediction (New Dataset)")
plt.xlabel('Years Of Exprience')
plt.ylabel("Salary")
plt.show()
```

From the above graph we can see that observations are in blue color where as regression line is in red color. As we can see that most of the observations are close to the regression line. Hence we can tell that simple regression model is a good model and able to make good predictions.
