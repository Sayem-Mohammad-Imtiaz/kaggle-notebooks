```python
df = pd.read_csv('../input/groceries/groceries - groceries.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Item(s)'], inplace=True,axis=1)
df.head()
```

```python
basket = []

for i in range(0,9834):
    basket.append([df.iloc[i,j] for j in range(0,32) if str(df.iloc[i,j]) != 'nan'])
basket
```

```python
from apyori import apriori
```

```python
rules = apriori(basket, min_support = 0.003, min_confidence = 0.2,min_lift = 3,min_length =2)
rules = list(rules)
len(rules)
```

```python
rules[0]
```

```python
rules[0].ordered_statistics[0].items_base
```

```python
rules[0].ordered_statistics[0].items_add
```

```python
rules[0].ordered_statistics[0].confidence
```

```python
result_df = pd.DataFrame(columns=['items_base','item_add','confidence'])
for i in range(len(rules)):
    items_base = rules[i].ordered_statistics[0].items_base
    item_add = rules[i].ordered_statistics[0].items_add
    confidence = rules[i].ordered_statistics[0].confidence
    
    result_df.loc[i] = [items_base,item_add,confidence]
```
