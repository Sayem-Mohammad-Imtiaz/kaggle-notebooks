```python
def detect_outliers (df,variable):
    outline_index = []
    
    for c in variable:
        Q1 = np.percentile(df[c],25)
        Q3 = np.percentile(df[c],75)
        IQR = Q3-Q1
        outlier_list = df[(df[c]<Q1-IQR) | (df[c]>Q3+IQR)].index
        outline_index.extend(outlier_list)
    return outline_index
```

```python
kc_house_data.loc[detect_outliers(kc_house_data,["sqft_living","grade","sqft_above","sqft_living15","bathrooms"])]
```

```python
kc_house_data_2 = kc_house_data.drop(detect_outliers(kc_house_data,["sqft_living","grade","sqft_above","sqft_living15","bathrooms"]),axis = 0).reset_index(drop = True)
```

# NORMALİZE

```python
kc_house_data_2['price'] = kc_house_data_2['price']/max( kc_house_data_2['price'])
kc_house_data_2['bathrooms'] = kc_house_data_2['bathrooms']/max( kc_house_data_2['bathrooms'])
kc_house_data_2['sqft_living'] = kc_house_data_2['sqft_living']/max( kc_house_data_2['sqft_living'])
kc_house_data_2['grade'] = kc_house_data_2['grade']/max( kc_house_data_2['grade'])
kc_house_data_2['sqft_above'] = kc_house_data_2['sqft_above']/max( kc_house_data_2['sqft_above'])
kc_house_data_2['sqft_living15'] = kc_house_data_2['sqft_living15']/max( kc_house_data_2['sqft_living15'])
```
