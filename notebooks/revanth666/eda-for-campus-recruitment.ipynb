# Exploratory Data Analysis

```python
df = pd.DataFrame(data)
df.head(2)
```

```python
rcParams['figure.figsize'] = 10,7
values = df['gender'].value_counts().values
index = df['gender'].value_counts().index
color = ['cyan','yellow']
plt.pie(values,labels = index,colors = color,autopct="%1.1f")
plt.title('Gender Classification')
plt.legend(loc = 'upper right')
plt.show()
```

```python
rcParams['figure.figsize'] = 10,5
sb.countplot(df['ssc_b'],palette=['c','b'])
plt.title('Secondary school board classification')
plt.show()
```

```python
rcParams['figure.figsize'] = 10,5
sb.countplot(df['hsc_b'],palette=['c','r'])
plt.title('Higher secondary school board classification')
plt.show()
```

```python
rcParams['figure.figsize'] = 10,7
values = df['hsc_s'].value_counts().values
index = df['hsc_s'].value_counts().index
color = ['cyan','blue','grey']
explode = [0,0,0.1]
plt.pie(values,labels = index,colors = color,autopct="%1.2f",explode = explode)
plt.title('Specialization in High school')
plt.legend(loc = 'upper right')
plt.show()
```

```python
rcParams['figure.figsize'] = 10,7
values = df['degree_t'].value_counts().values
index = df['degree_t'].value_counts().index
color = ['yellow','red','grey']
explode = [0,0,0.1]
plt.pie(values,labels = index,colors = color,autopct="%1.2f",explode = explode)
plt.title('Field of Degree')
plt.legend(loc = 'upper right')
plt.show()
```

```python
rcParams['figure.figsize'] = 10,7
values = df['workex'].value_counts().values
index = df['workex'].value_counts().index
color = ['lightskyblue','fuchsia']
plt.pie(values,labels = index,colors = color,autopct="%1.1f")
plt.title('Work Experience')
plt.legend(loc = 'upper right')
plt.show()
```

```python
rcParams['figure.figsize'] = 10,7
values = df['specialisation'].value_counts().values
index = df['specialisation'].value_counts().index
color = ['cyan','gold']
plt.pie(values,labels = index,colors = color,autopct="%1.1f")
plt.title('Specialisation')
plt.legend(loc = 'upper right')
plt.show()
```

```python
rcParams['figure.figsize'] = 10,7
values = df['status'].value_counts().values
index = df['status'].value_counts().index
color = ['lime','orange']
plt.pie(values,labels = index,colors = color,autopct="%1.1f")
plt.title('Placement status')
plt.legend(loc = 'upper right')
plt.show()
```
