**Näitab mängude ülemaailmseid müüginumbreid alates aastast 2000. Suurim on 2006 aasta Wii Sports 82,8mln.**

```python
df = pd.read_csv('../input/vgsales.csv')
df = df[(df["Year"] >= 2000) & (df["Year"] <= 2017) & (df["Global_Sales"] < 5)]
df.Global_Sales.plot.hist(grid = False, rwidth = 0.95, bins = 100);

```

**See graafik on iga programmeerija tulevik. Eksponentsiaalselt vähenev võimalus oma ideega läbi lüüa ja suurt raha teenida. 90% mängudest teenivad 0-20.000$ ehk ei jõua väga kaugele.**

```python
df = df[(df["Year"] >= 2000) & (df["Year"] <= 2017) & (df["Global_Sales"] >= 3.0) & (df["Global_Sales"] <= 50.0)]
df.plot.scatter("NA_Sales", "JP_Sales", alpha = 0.3);

```

**Graafik võrdlemaks, kuidas erinevate mängude müüginumbrid erinevad Põhja-Ameerikas ja Jaapanis.**

```python
tulp = pd.crosstab(df.Platform,df.Rank)
tulbasum = tulp.sum(axis=1).sort_values(ascending = True)
plt.figure(figsize=(12,6))
sns.barplot(y = tulbasum.index, x = tulbasum.values*40)
plt.show()
```

**Graafik näitab, kui palju müüdi mänge teatud mängukonsoolidele. Huvitav on tähelepanek, et kui varem oli Xbox PS3-st üle, siis nüüd on PS4 populaarsust tagasi saanud.**
