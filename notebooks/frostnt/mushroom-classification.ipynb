```python
def OHE(df, col):
    df = df.copy()
    dummy = pd.get_dummies(df[col], prefix=col, drop_first=True)
    df = pd.concat([df, dummy], axis=1)
    df = df.drop(col, axis=1)
    return df
```

To many object columns

```python
def preprocessInputs(df):
    # Handling the object columns
    df = df.copy()
    cols = df.columns
    for i in cols:
        df = OHE(df, i)
    
    # Splitting into X and y
    y = df['class_p']
    X = df.drop('class_p', axis=1)
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42, shuffle=True)
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = preprocessInputs(df)
```

# Model Creation

I feel like random forest would be a good model choice for this data set, so lets try using that

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)

# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
