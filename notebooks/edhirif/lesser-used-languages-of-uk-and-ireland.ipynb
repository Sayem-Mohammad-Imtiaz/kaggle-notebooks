These five categories are defined on the UNESCO '[Atlas of the World's Languages in Danger][1]' website.
I always like to start with an overview of the data:
  [1]: http://www.unesco.org/languages-atlas/

```python
df['Number of speakers'].groupby(df['Degree of endangerment']).describe()
```

I can see there are some problems with this data. The UNESCO website defines the 'Extinct' catagory as "there are no speakers left>>included in the Atlas if presumably extinct since the 1950s". This is pretty clear - extinct means zero speakers but five of the languages listed as extinct in the dataset have speaker numbers greater than zero, one of them has about twelve thousand speakers. I think some caution should be used when analysing this data.

I wanted to focus on the lesser used languages of Britain and Ireland as I am more familiar with these.

```python
df = df.drop(df.columns[[0, 2, 3, 4, 6, 8, 9, 12, 13]], axis=1)
df = df.rename(columns={'Name in English' : 'name', 'Country codes alpha 3': 'country',
                        'Degree of endangerment': 'endangerment', 'Number of speakers': 'speakers',
                        'Description of the location': 'location'})
df = df[df['country'].isin(['IRL', 'GBR', 'IRL, GBR', 'GBR, IRL'])]
dfsorted = df.drop(df.columns[[1,4,5]], axis=1)
dfsorted = dfsorted.sort_values('speakers',axis=0, ascending=False)
print(dfsorted)
```

The 2011 UK Census gives the number of Manx speakers as about 1,800 and the number of Cornish speakers as about 500 but this data is over five years old so is unlikely to be accurate now. Another problem with these numbers is-what does it mean to speak a language? Does it mean the person should be a native speaker, or can it include people who became fluent in later life, and what about people who have some level of fluency but are not at native speaker level. The Cornish language had to be reconstructed because the last native speaker died before audio recording equipment became available, she died in 1777. So does that mean that language is extinct and can never be brought back?
The above gives the number of Irish speakers as just over forty thousand (there are at least three times as many Polish speaking people living in Ireland) but I have seen the number of Irish speakers quoted as over one million. Politicians, the media and other commentators in Ireland will use a strict or non-strict definition of 'speaker' depending on which suits their political agenda.
There is also the question of what is a language and what is a dialect - many people, including the majority of people who speak Scots in Scotland do not consider Scots to be a language - see [here][1] for a report by the Scottish government. 
The topic of lesser used or minority languages is often politically sensitive, it is absolutely worth studying and it is important for us as a species to maintain our cultural diversity but I would treat the data in this dataset with some caution.
