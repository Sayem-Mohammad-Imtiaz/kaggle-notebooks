Import data, define a variable and a function - both of which will be used to get age

```python
import pandas as pd
import datetime as dt

end = pd.to_datetime('2016-08-21').date()
def correct_age(x):
    if  x < 0:
        x = x + 100
    return x

df = pd.read_csv('../input/athletes.csv', parse_dates=[4])
```

Add a column with total number of medals won; add a column with age

```python
df['medals'] = df['gold'] + df['silver'] + df['bronze']
df['age'] = end - pd.to_datetime(df.dob)
df['age'] = df['age'].astype('<m8[Y]')
df['age'] = df.age.apply(correct_age)
```

check new columns exist

```python
df.head()
```

create new dataframes for male and female  athletes who won at least one medal

```python
df_m = df.copy(deep=True)
df_m = df_m.drop(df_m[df_m.sex!='male'].index)
df_m_nz = df_m.copy(deep=True)
df_m_nz = df_m_nz.drop(df_m_nz[df_m_nz.medals==0].index) # male athletes who won at least one medal
df_m_z = df_m.copy(deep=True)
df_m_z = df_m_z.drop(df_m_z[df_m_z.medals!=0].index) # male athletes who did not win any medals

df_f = df.copy(deep=True)
df_f = df_f.drop(df_f[df_f.sex!='female'].index)
df_f_nz = df_f.copy(deep=True)
df_f_nz = df_f_nz.drop(df_f_nz[df_f_nz.medals==0].index) # female athletes who won at least one medal
df_f_z = df_f.copy(deep=True)
df_f_z = df_f_z.drop(df_f_z[df_f_z.medals!=0].index) # female athletes who did not win any medals
```

what percentage of male athletes won at least one medal?

```python
total_male_athletes = len(df_m.index)
num_male_medal = len(df_m_nz.index)
percent_m_winners = round((num_male_medal/total_male_athletes)*100,1)
percent_m_winners
```

what percentage of female athletes won at least one medal?

```python
total_female_athletes = len(df_f.index)
num_female_medal = len(df_f_nz.index)
percent_f_winners = round((num_female_medal/total_female_athletes)*100,1)
percent_f_winners
```

if you are a male athlete what sports give you the best chance of winning multiple medals?

```python
top_m = df_m[df_m.medals>=3]
top_m_sports = top_m.groupby('sport').size()
print(top_m_sports.sort_values(ascending=False))
```

```python
if you are a female athlete what sports give you the best chance of winning multiple medals?
```

```python
top_f = df_f[df_f.medals>=3]
top_f_sports = top_f.groupby('sport').size()
print(top_f_sports.sort_values(ascending=False))
```

The above shows that for both male and female athletes, aquatics is the best sport to win 3 or more medals.

is there a difference in the mean age of male atheletes, competing in aquatics, who won
medals and those who did not win any medals

```python
sports_m_z_age = df_m_z['age'].groupby(df_m_z['sport'])
sports_m_nz_age = df_m_nz['age'].groupby(df_m_nz['sport'])
sports_m_z_age_mean = sports_m_z_age.mean()
sports_m_nz_age_mean = sports_m_nz_age.mean()
print(sports_m_z_age_mean['aquatics'])
print(sports_m_nz_age_mean['aquatics'])
```

For male athletes competing in aquatic sports it seems that experience slightly outperforms youth - so the average age of those who won medals is slightly higher than those who did not

Repeating the above for female athletes

```python
sports_f_z_age = df_f_z['age'].groupby(df_f_z['sport'])
sports_f_nz_age = df_f_nz['age'].groupby(df_f_nz['sport'])
sports_f_z_age_mean = sports_f_z_age.mean()
sports_f_nz_age_mean = sports_f_nz_age.mean()
print(sports_f_z_age_mean['aquatics'])
print(sports_f_nz_age_mean['aquatics'])
```

The same is true for female athletes, also the average age for female athletes in aquatics is 1 or 2 years less than for males.

Repeating the above for athletics, first men

```python
print(sports_m_z_age_mean['athletics'])
print(sports_m_nz_age_mean['athletics'])
```

The average age of male athletics medal winners is higher than for aquatics and there is no significant difference between the average age of medal winners and those who did not win any medals

Repeating for female athletes

```python
print(sports_f_z_age_mean['athletics'])
print(sports_f_nz_age_mean['athletics'])
```

The same is true for female athletes, also there is no significant difference between the average ages of female and male athletes competing in athletics

Lastly, the same for gymnastics, first males

```python
print(sports_m_z_age_mean['gymnastics'])
print(sports_m_nz_age_mean['gymnastics'])
```

Then females

```python
print(sports_f_z_age_mean['gymnastics'])
print(sports_f_nz_age_mean['gymnastics'])
```

Again there is a significant difference in average ages for females and males, in gymnastics the female medal winners are significantly younger than the male medal winners. But there is no significant difference in average ages between those who won medals and those who did not.
