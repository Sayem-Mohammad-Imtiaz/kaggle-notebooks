**What is Red Shift?**

As objects move away from us the light travelling back to us gets 'stretched'. This stretching of the light causes the spectra to be shifted from their normal appearance towards the red end of the visible spectrum - red shift, it is a kind of Doppler shift. If an object were coming towards us it would be blue shifted. The cool thing about red shits is they give us information about the distance to objects. Red shift increases as the velocity increases. As Hubble (the astronomer not the telescope) demonstrated the further away an object is the faster it moves away from us, this is due to the expansion of the universe. The following boxplots for the three classes of object demonstarte that Quasars have the highest red shifts so they are further away, galaxies are closer and stars are closer still. 

```python
sns.boxplot( x=df["class"], y=df["redshift"] );
```

Information from stars, galaxies and quasars reaches us mostly in the form of electromagnetic radiation, for example visible light, radio waves, x-rays an so on. This information travels to us at a finite speed the speed of light, although it is fast - at this speed you could travel the 93 million miles to the sun in about 8 minutes, the distant universe is so far away it takes a very long time for the information to reach us. So we  see distant objects like quasars not as they are now but as they were billions of years ago. Since we mostly see quasars as distant not nearby objects we can say that the early universe contained more quasars while the universe now contains more galaxies.
