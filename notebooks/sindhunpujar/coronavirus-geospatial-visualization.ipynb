```python
fire_map = folium.Map([20.9517,85.0985],
                  zoom_start=8)
fire_map
```

```python
from branca.element import Figure
fig=Figure(width=550,height=350)
```

```python
fig2=Figure(width=550,height=350)
fire_map2=folium.Map(location=[22.9868,87.8550])
fig2.add_child(fire_map2)
folium.TileLayer('Stamen Terrain').add_to(fire_map2)
folium.TileLayer('Stamen Toner').add_to(fire_map2)
folium.TileLayer('Stamen Water Color').add_to(fire_map2)
folium.TileLayer('cartodbpositron').add_to(fire_map2)
folium.TileLayer('cartodbdark_matter').add_to(fire_map2)
folium.LayerControl().add_to(fire_map2)
fire_map2
```

```python
fire_map1=folium.Map(width=550,height=350,location=[22.9868,87.8550],zoom_start=11,min_zoom=8,max_zoom=14)
fig.add_child(fire_map1)
fire_map1
```

```python
# Creating Basemap
fig3=Figure(width=550,height=350)
fire_map3=folium.Map(location=[20.9517,85.0985],tiles='cartodbpositron',zoom_start=11)
fig3.add_child(fire_map3)

#Adding markers to the map
folium.Marker(location=[25.0961,85.3131],popup='Default Location1',tooltip='Click here to see Popup').add_to(fire_map3)
folium.Marker(location=[20.9517,85.0985],popup='<strong>Location3</strong>',tooltip='<strong>Click here to see Popup</strong>').add_to(fire_map3)
folium.Marker(location=[22.9868,87.8550],popup='<h3 style="color:blue;">Location2</h3>',tooltip='<strong>Click here to see Popup</strong>').add_to(fire_map3)
fire_map3
```

```python
# Creating Basemap
fig4=Figure(height=350,width=550)
fire_map4=folium.Map(location=[20.9517,85.0985],tiles='Stamen Terrain',zoom_start=11)
fig4.add_child(fire_map4)

# Adding Custom Markers
folium.Marker(location=[25.0961,85.3131],popup='Custom Marker 1',tooltip='<strong>Click here to see Popup</strong>',icon=folium.Icon(color='red',icon='none')).add_to(fire_map4)
folium.Marker(location=[20.9517,85.0985],popup='Custom Marker 2',tooltip='<strong>Click here to see Popup</strong>',icon=folium.Icon(color='green',prefix='glyphicon',icon='off')).add_to(fire_map4)
folium.Marker(location=[22.9868,87.8550],popup='Custom Marker 3',tooltip='<strong>Click here to see Popup</strong>',icon=folium.Icon(color='purple',prefix='fa',icon='anchor')).add_to(fire_map4)

fire_map4
```
