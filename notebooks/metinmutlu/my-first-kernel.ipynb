```python
df = pd.read_csv("/kaggle/input/bitcoin-historical-data/bitstampUSD_1-min_data_2012-01-01_to_2020-09-14.csv")
df.head()
```

```python
df.info() ## info about types of columns
```

```python
t_df = df.isna().sum() ## number of missing values
t_df
```

```python
df.shape ## total number of rows and columns
```

```python
df[df.isna().any(axis=1)] ## row with missing values
```

```python
## percentage of missing values
t_df = (t_df/df.shape[0])*100
t_df
```

```python
## convert unix to normal time format
df["Timestamp"] = pd.to_datetime(df["Timestamp"],unit="s")
df.head()
```

```python
df.Open.sample(20)
```

```python
#fill NaN values with median values 
# first need to find median values of columns
temp = df[pd.notna(df["Open"])]
temp.info()
```

```python
temp.shape
```

```python
med_temp = np.median(temp["Open"])
med_temp
```

```python
df["Open"].plot(kind="hist")
```

```python
# as seen above median values are applicable
# fill NaN values in Open, High, Low and Close
df["Open"].fillna(med_temp, inplace=True)
df["High"].fillna(med_temp, inplace=True)
df["Low"].fillna(med_temp, inplace=True)
df["Close"].fillna(med_temp, inplace=True)
```

```python
# fill other NaN values with 0
df["Volume_(BTC)"].fillna(0, inplace=True)
df["Volume_(Currency)"].fillna(0, inplace=True)
df["Weighted_Price"].fillna(0, inplace=True)
```

```python
df.isna().sum()
```

```python
df.corr()
```

```python
## Correlation graph
f,ax = plt.subplots(figsize=(10, 10))
sns.heatmap(df.corr(), annot=True, linewidth=1, fmt=".2f", center=0, ax=ax, cmap="YlGnBu")
plt.show()
```

```python
## lineplot
df["Open"].plot(kind="line", label="Open", alpha=0.5, color="red", linestyle=":", grid=True, linewidth=1)
df["Close"].plot(label="Close", alpha=0.5, color="blue", linestyle="-.", grid=True, linewidth=1)
plt.legend(loc="upper left")
plt.xlabel("x")
plt.ylabel("y")
plt.title("PRICE")
```
