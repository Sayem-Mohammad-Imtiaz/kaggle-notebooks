```python
dataset = pd.read_csv('/kaggle/input/weather-dataset-rattle-package/weatherAUS.csv')
```

```python
dataset
```

```python
x = dataset.iloc[:,[2,3,4,13,14,15,16,17,18,19,20]]
print(x)
y = dataset.iloc[:, 22]
y = pd.get_dummies(y, drop_first = True)
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.impute import SimpleImputer

my_imputer = SimpleImputer()
x_train= my_imputer.fit_transform(x_train)
x_test= my_imputer.fit_transform(x_test)
y_train = my_imputer.fit_transform(np.array(y_train).reshape(-1,1))
y_test = my_imputer.fit_transform(np.array(y_test).reshape(-1,1))
```

```python
print(y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF = RandomForestClassifier(n_estimators = 150)
RF.fit(x_train, y_train.ravel())
```

```python
ypred = RF.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, ypred)
print('Accuracy:', acc*100, '%')
```

```python
from xgboost import XGBClassifier

xgb = XGBClassifier()
xgb.fit(x_train, y_train.ravel())
```

```python
yp = xgb.predict(x_test)

acc2 = accuracy_score(y_test, yp)
print('Accuracy:', acc2*100, '%')
```
