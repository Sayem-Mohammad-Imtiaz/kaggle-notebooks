All columns have data type as 'Object'. Let's check data types using 'df.dtypes' and then convert the data types of numeric columns before we move on.

```python
# Covert Salary Range From data type to integer
NY= NY.astype({'Salary Range From':'int32'})
```

Sucess! Let's repeat for Salary Range To.

```python
# Covert Salary Range To data type to integer
NY= NY.astype({'Salary Range To':'int32'})
```

```python
# Check
print(NY["Salary Range To"].head())
```

Let's see the average salary ranges for each agency.

```python
# Grouping
NY.groupby('Agency')['Salary Range From','Salary Range To'].mean()
```

How about multiple statistics? Source: https://www.shanelynn.ie/summarising-aggregation-and-grouping-data-in-python-pandas/

```python
# Grouping with multiple stats
NY.groupby('Agency').agg(
    {
         'Salary Range From':"mean",
         'Salary Range To':"mean",    
         '# Of Positions': "count"  
    }
)
```

I don't like this output. Can we sort this output? 

```python
NY_groupby = NY.groupby('Agency').agg(
    {
         'Salary Range From':"mean",
         'Salary Range To':"mean",    
         '# Of Positions': "count"  
    }
)
NY_Sort = NY_groupby.sort_values(['# Of Positions'],ascending=False)
print(NY_Sort)
```

So, 'Department of Environment Protection' has the maximum number of job openings. Who has the highest average salary range to?

```python
NY_Sort2 = NY_groupby.sort_values(['Salary Range To'],ascending=False)
print(NY_Sort2)
```

We now know that 'NYC EMPLOYEES RETIREMENT SYS ' has the highest average salary of $ 136,976 with 52 open positions.

```python
# Groupby multiple columns
NY.groupby(['Agency','Job Category']).agg(
    {
         'Salary Range From':"mean",
         'Salary Range To':"mean",    
         '# Of Positions': "count"  
    }
)
```

Obviously, now we are interested in which Agency and Job Category within has highest openings!

```python
NY_multiple_groupby = NY.groupby(['Agency','Job Category']).agg(
    {
         'Salary Range From':"mean",
         'Salary Range To':"mean",    
         '# Of Positions': "count"  
    }
)
NY_Sort3 = NY_multiple_groupby.sort_values(['# Of Positions'],ascending=False)
print(NY_Sort3)
```

Unsurprisingly, 'DEPT OF ENVIRONMENT PROTECTION' is the agency with maximum positions within the job category 'Engineering, Architecture, & Planning' with 213 positions and average 'salary range to' of $704,225. What do I have to do for this salary!

Now I am interested in which 'Business Title' has higher salaries in general?

```python
NY_groupby_2 = NY.groupby('Business Title').agg(
    {
         'Salary Range From':"mean",
         'Salary Range To':"mean",    
         '# Of Positions': "count"  
    }
)
NY_Sort4 = NY_groupby_2.sort_values(['Salary Range To'],ascending=False)
print(NY_Sort4)
```

Wow! The title of 'executive Vice President for Operations' has a max(only 2 positions,so it's max) salary of $234,402.
