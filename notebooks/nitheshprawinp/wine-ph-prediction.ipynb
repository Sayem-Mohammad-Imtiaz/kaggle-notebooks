Importing necessary libraries 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split as tts
from sklearn.linear_model import LinearRegression as li_reg
from sklearn.metrics import accuracy_score as acc_scr
```

```python
df = pd.read_csv('../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
```

```python
def bound_cal (data):
    
    q1 = np.quantile(data,.25)
    q3 = np.quantile(data,.75)
    iqr = q3-q1
    lb = q1-(1.5*iqr)
    hb = q3+(1.5*iqr)
    return lb,hb
```

```python
lb_tsd,hb_tsd=bound_cal(df['total sulfur dioxide'])
lb_fsd,hb_fsd=bound_cal(df['free sulfur dioxide'])
lb_cl,hb_cl=bound_cal(df['chlorides'])
```

```python
df.drop(df[df['total sulfur dioxide']>hb_tsd].index,axis=0,inplace=True)
df.drop(df[df['free sulfur dioxide']>hb_fsd].index,axis=0,inplace=True)
```

```python
df.drop(df[(df['chlorides']>hb_cl) | (df['chlorides']<lb_cl)].index,axis=0,inplace=True)
```

```python
mask=np.triu(df.corr())
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),mask=mask,annot=True)
```

Features with good correlation to pH:

* fixed acidity
* volatile acidity
* citric acid
* chlorides
* free sulfur dioxide
* density
* alcohol

```python
y = df['pH']
x = df[['fixed acidity','volatile acidity','citric acid','chlorides','free sulfur dioxide','density','alcohol']]
```

```python
x_train,x_test,y_train,y_test = tts(x,y,test_size=.23)
```

```python
model = li_reg().fit(x_train,y_train)
```

```python
y_hat=model.predict(x_test)
```

```python
plt.figure(figsize=(10,6))
plt.plot(y_hat)
```

```python
y_test=y_test.reset_index(drop=True)
```

```python
plt.plot(y_test)
plt.plot(y_hat)
```
