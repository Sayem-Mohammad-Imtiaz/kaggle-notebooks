```python
# Import necessary packages for exploratory analysis and data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Exploratory analysis
data = pd.read_csv('/kaggle/input/used-car-dataset-ford-and-mercedes/toyota.csv')

data.info()
data.describe()
data.head()
```

```python
# Visualizing price vs. mileage by model
sns.scatterplot(data = data, x = 'mileage', y = 'price', hue = 'model', palette = 'tab20')
```

```python
# Import necessary packages for model development
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
# Set random seed for reproducibility
rs = 420
```

```python
# Encoding features and assigning X variable
enc = OneHotEncoder()
X = enc.fit_transform(data.drop('price', axis = 1).copy())
```

```python
# Assigning y variable
y = data['price'].copy()
```

```python
# Split data into training and test subsets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = rs)
```

```python
# Model development
model = GradientBoostingRegressor(loss = 'huber', learning_rate = 0.65)
model.fit(X_train, y_train)
preds = model.predict(X_test)
    
r2_score(y_test, preds)
```
