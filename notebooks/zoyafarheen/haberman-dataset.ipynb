```python
#CDF: Cumulative Distribution Function, returns the probability of a value less than or equal to a given outcome
class_1 =haberman.loc[haberman["Survival-status"]==1]
class_2 =haberman.loc[haberman["Survival-status"]==2]
```

```python
#Age feature corresponding to class 1
counts,bin_edges =np.histogram(class_1['Age'],bins=10,density =True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf ,label = 'cdf')
plt.xlabel('Age')
plt.title("[CDF:PLOT 1.1] Age feature corresponding to class1")
plt.legend()
plt.show()
```

* it is clear that patients having Age between 50 to 60 have greater rate of survival
* Almost 90% patient who survived have the age less than or equal to 70

```python
#Age failure corresponding to class 2
counts,bin_edges =np.histogram(class_2['Age'],bins=10,density =True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf ,label = 'cdf')
plt.xlabel('Age')
plt.title("[CDF:PLOT 1.1] Age feature corresponding to class 2")
plt.legend()
plt.show()
```

*  it is observed that Around 74% of patient who dies within 5 years have an age less than or equal to 60 years

```python
#Operation year corresponding to class 1
counts,bin_edges =np.histogram(class_1['Operation-year'],bins=10,density =True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf ,label = 'cdf')
plt.xlabel('Operation-year')
plt.title("[CDF:PLOT 1.1] Operation year corresponding to class1")
plt.legend()
plt.show()
```

* it is observed that 60% patients are undergoing surgery before the year 1964 and survived

```python
#Operation year corresponding to class 2
counts,bin_edges =np.histogram(class_2['Operation-year'],bins=10,density =True)

pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf ,label = 'cdf')
plt.xlabel('Operation-year')
plt.title("[CDF:PLOT 1.1] Operation year corresponding to class2")
plt.legend()
plt.show()
```

```python
#Axil Nodes corresponding to class a
counts,bin_edges =np.histogram(class_1['Axil-Nodes'],bins=10,density =True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],cdf ,label = 'cdf')
plt.xlabel('Axil-Nodes')
plt.title("[CDF:PLOT 1.1] Axil Nodes corresponding to class1")
plt.legend()
plt.show()
```

 # 1. KNN algorithm 

```python
x = haberman.iloc[:,:2]
y = haberman["Survival-status"]
```

```python
from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=0)  
  
#feature Scaling  
from sklearn.preprocessing import StandardScaler    
st_x= StandardScaler()    
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test)
```

```python
from sklearn.model_selection import train_test_split  
from sklearn.preprocessing import StandardScaler 
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn import metrics
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=0)   
#feature Scaling   
st_x= StandardScaler()    
x_train= st_x.fit_transform(x_train)    
x_test= st_x.transform(x_test) 
classifier= KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2 )  
classifier.fit(x_train, y_train)  
```

```python
from sklearn.model_selection import cross_validate
from sklearn.model_selection import cross_val_score
print(cross_val_score(classifier,x,y,cv=5))
```

```python
#Train
y_pred= classifier.predict(x_train) 
score= metrics.accuracy_score(y_train,y_pred)
print("Your Model Accuracy is", score)
```

```python
#Test
y_pred= classifier.predict(x_test)  
score2= metrics.accuracy_score(y_test,y_pred)
print(cross_val_score(classifier,x,y,cv=5))
print("Your Model Accuracy is", score2)
```

```python
#Visulaizing the trianing set result  
from matplotlib.colors import ListedColormap  
x_set, y_set = x_train, y_train  
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step  =0.01),  
np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))  
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),  
alpha = 0.75, cmap = ListedColormap(('red','green' )))  
plt.xlim(x1.min(), x1.max())  
plt.ylim(x2.min(), x2.max())  
for i, j in enumerate(np.unique(y_set)):  
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],  
        c = ListedColormap(('green', 'red'))(i), label = j)  
plt.title('K-NN Algorithm (Training set)')  
plt.xlabel('Survival-status')  
plt.ylabel('Age')  
plt.legend()  
plt.show()  
```

```python
#Visualizing the test set result  
from matplotlib.colors import ListedColormap  
x_set, y_set = x_test, y_test  
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step  =0.01),  
np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))  
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),  
alpha = 0.75, cmap = ListedColormap(('green','red' )))  
plt.xlim(x1.min(), x1.max())  
plt.ylim(x2.min(), x2.max())  
for i, j in enumerate(np.unique(y_set)):  
        plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],  
        c = ListedColormap(('green', 'red'))(i), label = j)  
plt.title('K-NN algorithm(Test set)')  
plt.xlabel('Survial-status')  
plt.ylabel('Age')  
plt.legend()  
plt.show()  
```

# 2.Support vector

```python
#Extracting Independent and dependent Variable  
x1 = haberman.iloc[:,:2]
y1 = haberman["Survival-status"]
```

```python
from sklearn.model_selection import train_test_split  
from sklearn.preprocessing import StandardScaler
x_train, x_test, y_train, y_test= train_test_split(x1, y1, test_size= 0.25, random_state=0)  
#feature Scaling      
st_x= StandardScaler()    
x_train= st_x.fit_transform(x_train)   
```

```python
from sklearn.svm import SVC # "Support vector classifier"
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split   
classifier = SVC(kernel='linear', random_state=0)  
classifier.fit(x_train, y_train)
x_train, x_test, y_train, y_test= train_test_split(x1, y1, test_size= 0.25, random_state=0)  
```

```python
#Train
y_pred= classifier.predict(x_train)  
cm= confusion_matrix(y_train, y_pred)
score3= metrics.accuracy_score(y_train,y_pred)
print("Your Model Accuracy is", score3)
print(cm)
print(cross_val_score(classifier,x1,y1,cv=5))
```

```python
#Test 
y_pred= classifier.predict(x_test)  
cm= confusion_matrix(y_test, y_pred)
score4= metrics.accuracy_score(y_test,y_pred)
print("Your Model Accuracy is", score4)
print(cm)
```

# 3.Decision Tree

```python
x2 = haberman.iloc[:,:2]
y2= haberman["Survival-status"]
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x2, y2, test_size = 0.25, random_state = 0)  
  
# Feature Scaling  
from sklearn.preprocessing import StandardScaler  
sc = StandardScaler()  
x_train = sc.fit_transform(x_train)  
x_test = sc.transform(x_test) 
classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)  
classifier.fit(x_train, y_train) 
```

```python
#Train 
y_pred= classifier.predict(x_train)  
score5= metrics.accuracy_score(y_train,y_pred)
print("Your Model Accuracy is", score5)
print(cross_val_score(classifier,x2,y2,cv=5))
```

```python
#Test
y_pred= classifier.predict(x_test)  
score6 = metrics.accuracy_score(y_test,y_pred)
print("Your Model Accuracy is", score6)
```

# 4.Random Forest 

```python
from sklearn.ensemble import RandomForestClassifier 
x3 = haberman.iloc[:,:2]
y3 = haberman["Survival-status"]
# Splitting the dataset into training and test set.  
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x3, y3, test_size= 0.25, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier  
classifier= RandomForestClassifier(n_estimators= 16, criterion="entropy")  
classifier.fit(x_train, y_train)  
```

```python
#Predicting the test set result  
#Train
y_pred= classifier.predict(x_train)  
score7 = metrics.accuracy_score(y_train, y_pred)
print("Accuracy:",score7)
print(cross_val_score(classifier,x3,y3,cv=5))
```

```python
#Test
y_pred= classifier.predict(x_test)  
score8 = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:",score8)
cm3= confusion_matrix(y_test, y_pred)
print(cm3)
```
