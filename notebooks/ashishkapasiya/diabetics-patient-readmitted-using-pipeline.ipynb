```python
train['age'] = train['age'].replace({'[70-80)': 75, '[60-70)': 65, '[50-60)': 55, '[80-90)': 85, '[40-50)': 45, '[30-40)': 35, '[90-100)': 95, '[20-30)': 25, '[10-20)': 15, '[0-10)': 5})
```

```python
train['number_patients'] = train['number_outpatient'] + train['number_inpatient']
train.drop(['number_outpatient', 'number_inpatient'], axis = 1, inplace = True)
```

```python
df = train[['metformin','repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride','acetohexamide', 'glipizide', 
                'glyburide', 'tolbutamide','pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone',
                'tolazamide', 'glyburide-metformin', 'glipizide-metformin','glimepiride-pioglitazone',
                'metformin-rosiglitazone','metformin-pioglitazone', "insulin"]]
```

```python
def med_count(df):
    df['Steady'] = np.sum(df == "Steady")
    df['No'] = np.sum(df == "No")
    df['Up'] = np.sum(df =='Up')
    df['Down'] = np.sum(df =='Down')
    return df
```

```python
df = df.apply(med_count, axis = 1)
df.head()
```

```python
df[['Steady', 'Up', "Down", 'No']].describe()
```

```python
for i in df[['Steady', 'Up', "Down", 'No']].columns:
    print(i, df[i].unique(), "\n")
```

```python
count_col = df[['Steady', 'Up', "Down", 'No']]
train = pd.concat([train,count_col], axis = 1)
```

```python
train['Steady'] = train['Steady'].replace(2, 1).replace(3, 1).replace(4, 1).replace(5, 1).replace(6, 1)
train['Up'] = train['Up'].replace(2, 1).replace(3, 1)
train['Down'] = train['Down'].replace(2, 1).replace(3, 1)
train['No'] = train['No'].replace(21, 1).replace(20, 1).replace(19, 1).replace(18, 1).replace(17, 1).replace(16, 1).replace(15, 1)

for i in train[['Steady', 'Up', "Down", 'No']].columns:
    print(i, train[i].unique(), "\n")
```

```python
train.drop(['metformin','repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride','acetohexamide', 'glipizide', 
                'glyburide', 'tolbutamide','pioglitazone', 'rosiglitazone', 'acarbose', 'miglitol', 'troglitazone',
                'tolazamide', 'glyburide-metformin', 'glipizide-metformin','glimepiride-pioglitazone',
                'metformin-rosiglitazone','metformin-pioglitazone', "insulin"], axis = 1, inplace = True)
```

```python
train.drop('No', axis = 1, inplace = True)
```

```python
train['change'] = train['change'].replace("No", 0).replace("Ch", 1)
train['diabetesMed'] = train['diabetesMed'].replace("No", 0).replace("Yes", 1)
```

```python
train['A1Cresult'] = train['A1Cresult'].replace(">7", 1).replace(">8", 1).replace("None", 0).replace("Norm", 2)
train['max_glu_serum'] = train['max_glu_serum'].replace(">200", 1).replace(">300", 1).replace("None", 0).replace("Norm", 2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(train.drop(columns=['readmitted']),
                                                 train['readmitted'],
                                                 test_size=0.3,
                                                random_state=8)
```

```python
trf1 = ColumnTransformer([
    ('cat_col',OneHotEncoder(sparse=False,handle_unknown='ignore'),[1,2,4,5,6])
],remainder='passthrough')

trf2 = LogisticRegression()
```

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2)
])
```

```python
pipe.fit(X_train,y_train)
y_pred = pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train, y_train, cv=5, scoring='accuracy').mean()
```
