```python
# Scatter Plot 
# x = EU Sales, y = NA Sales
data.plot(kind='scatter', x='EU_Sales', y='NA_Sales',alpha = 0.5,color = 'red',figsize=(10,10))
plt.xlabel('EU Sales')              # label = name of label
plt.ylabel('NA Sales')
plt.title('EU - NA Sales Scatter Plot')            # title = title of plot
```

A comparison of the level at which the relevant games are sold in the two regions is drawn through the scatter plot. Because of some games have very high sales levels, there are clusters of games with low sales levels.

İlgili oyunların iki bölgede hangi seviyede satıldığının karşılaştırmasını scatter plot aracılığıyla çizdirmiş bulunmaktayım. Fakat bazı oyunlar çok yüksek satış seviyelerine sahip olduğundan dolayı düşük satış seviyelerine sahip oyunlarda kümelenmeler bulunmaktadır.

Because drawing such graphs in data is meaningless at the moment, more meaningful graphs can be extracted by performing various filtering operations.

Verisetinde bu gibi grafiklerin çizdirilmesi şu an için anlamsız geldiğinden dolayı çeşitli filtreleme işlemleri yapılarak daha anlamlı grafikler çıkarılabilir.

```python
# 1 - Filtering Pandas data frame
x = data['EU_Sales']>2.47     # There are 100 games which have higher sales level than 2.47 million sales in Europe
data[x]
```

```python
x = data[(data['EU_Sales']>2.47) & (data['NA_Sales']>2.5)]    # There are 100 games which have higher sales level than 2.47 million sales in Europe
x
```

```python
# Scatter Plot 
# x = EU Sales, y = NA Sales
x.plot(kind='scatter', x='EU_Sales', y='NA_Sales',alpha = 0.5,color = 'red',figsize=(10,10))
plt.xlabel('EU Sales')              # label = name of label
plt.ylabel('NA Sales')
plt.title('EU - NA Sales Scatter Plot')            # title = title of plot
```

I created a new data frame called 'x' and took the sales levels in the intersection set of sales in the EU and NA regions, which I filtered into this data frame in. In this way, I created line and scatter plots that can be interpreted more comfortably.

'x' adlı yeni bir dataframe oluşturup bu dataframein içine filtreleme yaptığım EU ve NA bölgelerindeki satışların kesişim kümesinde bulunan satış seviyelerini aldım. Bu sayede derece daha rahat yorumlanabilen line ve scatter plotları oluşturdum.

```python
y = sales_data['Global_Sales']
y
```

```python
features = ['EU_Sales','NA_Sales','JP_Sales','Other_Sales']
X = x[features]
X
```

```python
from sklearn.tree import DecisionTreeRegressor
#specify the model. 
#For model reproducibility, set a numeric value for random_state when specifying the model
model = DecisionTreeRegressor(random_state = 4)
# Fit the model
model.fit(X, y)
```

```python
predictions = model.predict(X)
print(predictions)
```

```python
# Import the train_test_split function and uncomment
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split( X, y, random_state = 4)
```

```python
# Specify the model
model = DecisionTreeRegressor(random_state = 4)

# Fit model with the training data.
model.fit(train_X,train_y)
```

```python
# Predict with all validation observations
val_predictions = model.predict(val_X)
```

```python
val_predictions
```

```python
#validation predictions
print(val_predictions)
#actual prices from validation data
print(val_y)
```

```python
from sklearn.metrics import mean_absolute_error
val_mae = mean_absolute_error(val_y,val_predictions)

#validation_mae
print(val_mae)
```
