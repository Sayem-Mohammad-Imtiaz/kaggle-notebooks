# Unsupervised Learning

* We have not label in data and we dont know which value in which class.
* we just separate the data to the classes

* I will try to separate the countries about health level

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../input/unsupervised-learning-on-country-data/Country-data.csv")
df.head()
```

lets look at the connection between the two feature 

```python
plt.scatter(df["health"], df["child_mort"])
plt.xlabel("health")
plt.ylabel("child mortality")
plt.show()
```

* let's use KMeans model for clustering the data
* but first we have to choose k value 
* we can compare k vlaues and inertia values and we can choose the elbow point
* do this with for loop

```python
data = df.loc[:,["health","child_mort"]]

inertia_list = []

from sklearn.cluster import KMeans
for k in range(1,15):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data)
    inertia_list.append(kmeans.inertia_)
plt.plot(range(1,15), inertia_list, "-o", color="darkred")
plt.xlabel("k values")
plt.ylabel("inertia values")
plt.show()
```

* 2 is the elbow point for clustering

```python
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)
labels = kmeans.predict(data)

plt.scatter(data["health"], data["child_mort"], c = labels)
plt.show()
```

* ve didnt standardization the data 
* let's do standardization and implement all features into the model

* there is 2 class
* first class has the lower health level
* second class has the higher health level

```python
data2 = df.drop("country", axis=1)

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
standart = StandardScaler()
kmeans = KMeans(n_clusters = 2)
pipe = make_pipeline(standart, kmeans)
pipe.fit(data2)
labels = pipe.predict(data2)

data_classed = pd.DataFrame({"countries":df["country"], "class":labels})

print(data_classed[data_classed["class"] == 0])
print(data_classed[data_classed["class"] == 1])
```

* and we can cluster with hierarchical model
* firs create a dendrogram graph with using scipy

```python
from scipy.cluster.hierarchy import linkage, dendrogram
merg = linkage(data2.iloc[:35,:], method = "single")
dendrogram(merg, leaf_rotation=90, leaf_font_size=6)
plt.show()
```

* we put a threshold in longest vertical line but not choose the one which started from floor 
* after this  we count the point that coincide with threshold line 
* this is the cluster count

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

standart = StandardScaler()
hierarchy = AgglomerativeClustering(n_clusters=2, affinity="euclidean", linkage="ward")

pipe = make_pipeline(standart, hierarchy)
labels = pipe.fit_predict(data2)

data_hierarchy = pd.DataFrame({"country":df["country"], "class":labels})
print(data_hierarchy[data_hierarchy["class"]==0])
print(data_hierarchy[data_hierarchy["class"]==1])
```
