# Input

```python
df = pd.read_csv('/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv')
df_population = pd.read_csv('/kaggle/input/population-by-country-2020/population_by_country_2020.csv')
```

## Data Cleaning

### Filling up all NaN values with 'zero', as we all know the vaccination drive is quite new which has started almost in the begining of 2021 in most of the countries.

```python
df=df.fillna(0)
```

### Droppping 'England', 'Scotland', 'Wales', 'Northern Ireland' from country column as they are part of UK

```python
df = df[df.country.apply(lambda x: x not in ['England', 'Scotland', 'Wales', 'Northern Ireland'])]
```

```python
countries_list=list(df['country'].unique())
vaccine_list=list(df['vaccines'].unique())
vaccine_col = list(df['vaccines'])
country_col = list(df['country'])
people_vaccinated_col = list(df['people_vaccinated'])
```

```python
index_list = list()
for i in range(len(countries_list)):
    for j in range(len(country_col)):
        if(countries_list[i] == country_col[j]):
            index_list.append(j)
            break
            
vaccine_used_list=list()
for item in index_list:
    vaccine_used_list.append(vaccine_col[item])

vaccine_used_df = pd.DataFrame()
vaccine_used_df['Country'] = countries_list
vaccine_used_df['Vaccine_used'] = vaccine_used_list
```

## Which country has fully vaccinated more people at latest ?


```python
sum_people_vacc = df.groupby('country').sum()['daily_vaccinations']
```

```python
people_vaccinated_df = pd.DataFrame(data=sum_people_vacc)
people_vaccinated_df= people_vaccinated_df.reset_index()
people_vaccinated_df
```

### Top 10 Countries in terms of daily vaccinations
Made a list of top 10 countries who are doing great in terms of daily vaccinations

```python
top_10= people_vaccinated_df.sort_values(by='daily_vaccinations', ascending=False)[:10]
top_10
```

```python
top_10_countries = list(top_10['country'])
pop_country= list(df_population['Country (or dependency)'])
pop_pop = list(df_population['Population (2020)'])

pop_10_country = list()
pop_10_pop = list()
for i in range(len(top_10_countries)):
    for j in range(len(pop_country)):
        if(top_10_countries[i] == pop_country[j]):
                   pop_10_country.append(pop_country[j])
                   pop_10_pop.append(pop_pop[j])
df_top_10 = pd.DataFrame()
df_top_10['Country'] = pop_10_country
df_top_10['Population'] = pop_10_pop

df_top_10
```

## Number of people fully vaccinated for these top 10 countries.
Now we can dive in and look how many people are actually fully vaccinated for these countries !!

**Interesting** to note as there is no data available for **China** in terms of fully vaccinated people in this dataset.

```python
fully_vaccinated_people = list()
for item in top_10_countries:
    latest_no=df.groupby(by='country').get_group(item)['people_fully_vaccinated'].values[-1]
    fully_vaccinated_people.append(latest_no)
df_fully_vaccinated=pd.DataFrame()
df_fully_vaccinated['Country']= top_10_countries
df_fully_vaccinated['People_fully_vaccinated'] = fully_vaccinated_people

df_fully_vaccinated
```

## Graph Visual

### Number of people that are fully vaccinated  Versus  their country population at latest !

Interesting to note as there is no data available for China in terms of fully vaccinated people in this dataset.


```python
f, ax = plt.subplots(figsize=(15,15))

sns.set_color_codes("muted")
sns.barplot(x="Population", y='Country',label='Total Population' ,data=df_top_10, color='b')

sns.set_color_codes("pastel")
sns.barplot(y="Country", x='People_fully_vaccinated', label='Fully Vaccinated People',data=df_fully_vaccinated, color='b')

ax.legend(ncol=1, loc="lower right", frameon=True)
ax.set(ylabel="Countries",
       xlabel="Population in billions")
sns.despine(left=True, bottom=True)
```

* Another thing to notice , as we see from the graph for almost every country except Israel the no. of people that are fully vacccinated are actually quite less !!
