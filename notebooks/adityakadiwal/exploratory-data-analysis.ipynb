```python
# Third party packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Charting options
sns.set(context= "notebook", color_codes=True)
%matplotlib inline
```

```python
df = pd.read_csv("/kaggle/input/summer-products-and-sales-in-ecommerce-wish/summer-products-with-rating-and-performance_2020-08.csv")
df.head()
```

```python
df_prod = df[["title_orig", "price", "retail_price", "units_sold", "uses_ad_boosts", "rating", "rating_count", "rating_five_count","rating_four_count","rating_three_count","rating_two_count", "rating_one_count", "tags", "product_color", "shipping_option_price", "countries_shipped_to", "origin_country", "theme", "merchant_name"]]
df_merchant = df[["units_sold", "uses_ad_boosts", "merchant_title", "merchant_name", "merchant_info_subtitle", "merchant_rating_count", "merchant_rating", "merchant_id"]]
```

```python
plt.figure(figsize = (20,5))
sns.scatterplot(x = "rating_five_count", y = "retail_price", hue="uses_ad_boosts", data = df_prod, alpha = 0.7, palette="viridis");
```

- Looks like there are a lot of products that have more than 2000 five star ratings sold by merchants who do not use ad boosts more than ones who do.
- There are a lot of products who have a high **retail_price** but very few **five star ratings** at all. Don't buy these.

```python
sns.distplot(df_prod["rating_count"], bins=10);
```
