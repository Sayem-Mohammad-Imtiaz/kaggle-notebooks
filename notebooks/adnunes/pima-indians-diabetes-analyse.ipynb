```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import statsmodels.api as sm
```

```python
y = data['Outcome']
```

```python
X = data[['Glucose', 'BMI', 'Pregnancies','DiabetesPedigreeFunction','BloodPressure']]
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X,y, test_size=0.2, random_state=1610)
```

```python
X_train_constante = sm.add_constant(X_train)
```

```python
modelo_statsmodels = sm.OLS(y_train, X_train_constante, hasconst = True).fit()
print(modelo_statsmodels.summary())
```
