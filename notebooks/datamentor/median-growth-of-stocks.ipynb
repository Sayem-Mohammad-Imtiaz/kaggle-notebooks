```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import matplotlib.pyplot as plt

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv("/kaggle/input/nifty50-stock-market-data/TITAN.csv")
df.head()
```

Objective is to find the growth of a stock for the span of years 2000 to 2021. Here the growth is being computed for all the consecutive intervals with the years 2000 and 2021. Then we are taking the median of the growth values.

```python
# extract years from the dates
df['year'] = [i[:4] for i in df['Date']]

# convert 'year' column to int datatype
df['year'] = df['year'].astype('int')

# aggregate (median) closing price by years
df_aggr = df.groupby(['year'])['Close'].median().reset_index()

# function to compute growth for consecutive 3, 5 and 10 yearly intervals
def yr_growth(yr, price, freq):
    yr = list(yr)
    price = list(price)
    
    growth = []
    
    for i,y in enumerate(yr):
        if i+freq <= len(yr)-1:
            g = price[i+freq]/price[i]
            growth.append(g)
            
    return growth

# find three year growth values
three_yr_growth = yr_growth(df_aggr['year'], df_aggr['Close'], 3)

# find five year growth values
five_yr_growth = yr_growth(df_aggr['year'], df_aggr['Close'], 5)

# find seven year growth values
seven_yr_growth = yr_growth(df_aggr['year'], df_aggr['Close'], 7)

# find ten year growth values
ten_yr_growth = yr_growth(df_aggr['year'], df_aggr['Close'], 10)
```

```python
print("Average 3 year growth:",round(np.median(three_yr_growth),2))
plt.plot(three_yr_growth);
```

```python
print("Average 5 year growth:",round(np.median(five_yr_growth),2))
plt.plot(five_yr_growth);
```

```python
print("Average 7 year growth:",round(np.median(seven_yr_growth),2))
plt.plot(seven_yr_growth);
```

```python
print("Average 10 year growth:",round(np.median(ten_yr_growth),2))
plt.plot(ten_yr_growth);
```
