```python
import numpy as np  # linear algebra
import array as arr
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt #matplotlib.pyplot is a collection of command style functions that make matplotlib work like MATLAB. 
#Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, plots some lines in a plotting area, 
#decorates the plot with labels, etc.

import statistics
from statistics import stdev 


#Import these things to play with Violin Plot
import plotly.plotly as py
import plotly.figure_factory as ff
from plotly.graph_objs import graph_objs
from scipy import stats
import seaborn as sns
import csv


import os
print(os.listdir("../input"))

ds = pd.read_csv('../input/IRIS.csv') #In this line, I just import the dataset into Python. 
#For more information about the dataset: https://en.wikipedia.org/wiki/Iris_flower_data_set

print (ds.columns) #This line will print the headings of the table out 
print (len(ds.columns)) #This line will display the number of colummns out 
print (ds.head()) #display the first 5 records 

print (ds[['sepal_length','species']]) #This line is used to print the two columns out (the "sepal_length" and the "species")
#ds  #This line will print the entire dataset out. I commented it out because it is so lengthy.

ds.info()  #This line checking if there is any inconsistency in the dataset

ds.hist(edgecolor='red', linewidth=1.2)
fig=plt.gcf() #plt.gcf() allows us to get a reference to the current figure when using pyplot.
#GCF stands for Get Current Figure

fig.set_size_inches(12,10) #This is just the line to set the size for the graph 
plt.show() 
#See this one: http://queirozf.com/entries/matplotlib-pylab-pyplot-etc-what-s-the-different-between-these#plt-gcf

#Create a scatter graph of Setosa, Versicolor, and Virginica based on Sepal_leng, Sepal_width
fig = ds[ds.species=='Iris-setosa'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='orange', label='Setosa')
ds[ds.species=='Iris-versicolor'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='blue', label='Versicolor',ax=fig)
ds[ds.species=='Iris-virginica'].plot(kind='scatter',x='sepal_length',y='sepal_width',color='green', label='Virginica', ax=fig)

fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepal Length vs Width")

fig=plt.gcf()
fig.set_size_inches(10,10)
plt.show()

#Create a scatter graph of Setosa, Versicolor, and Virginica based on petal_leng, petal_width
fig = ds[ds.species=='Iris-setosa'].plot(kind='scatter',x='petal_length',y='petal_width',color='orange', label='Setosa')
ds[ds.species=='Iris-versicolor'].plot(kind='scatter',x='petal_length',y='petal_width',color='blue', label='Versicolor',ax=fig)
ds[ds.species=='Iris-virginica'].plot(kind='scatter',x='petal_length',y='petal_width',color='green', label='Virginica', ax=fig)

fig.set_xlabel("Petal Length")
fig.set_ylabel("Petal Width")
fig.set_title("Petal Length VS Width")

fig=plt.gcf()
fig.set_size_inches(10,10)
plt.show()


plt.figure(figsize=(20,10))
plt.subplot(2,2,1) 
sns.violinplot(x='species',y='petal_length',data=ds)

plt.subplot(2,2,2)
sns.violinplot(x='species',y='petal_width',data=ds)

plt.subplot(2,2,3)
sns.violinplot(x='species',y='sepal_length',data=ds)

plt.subplot(2,2,4)
sns.violinplot(x='species',y='sepal_width',data=ds)


filename = '../input/IRIS.csv'
with open(filename) as ds:
    reader = csv.reader(ds)
    header_row = next(reader)

    sepal_length = []
    
    
for row in reader:
        sepal_length.append(float(row[0]))
    
print("Setosa sepal length's data and Standard deviation: ")
print(sepal_length[0:50]);
print(statistics.stdev(sepal_length[0:50]));
    
print("Versicolor sepal length's data and Standard deviation: ")
print(sepal_length[50:100]);
print(statistics.stdev(sepal_length[50:100]));
    
print("Virginica sepal length's data and Standard deviation: ")
print(sepal_length[100:150]);
print(statistics.stdev(sepal_length[100:150]));

from pandas.tools.plotting import andrews_curves
andrews_curves(axis=1), "species")

```
