**Description**
In October of 2018, the Board of Supervisors passed and Mayor London N. Breed signed, the Ordinance 243-18 (http://bit.ly/2LSqS3U) to expand the extent to which women are represented in the public sphere, including within artwork, statues, street names, facilities, parks, and more. The representation of women in City-owned buildings includes buildings, conference rooms, clubhouses, museums, recreation centers, community rooms, auditoriums, terminals, departure halls, 7 staircases, rooms, and other places open to the public. The City Administrator's office was responsible for compiling a list of all city owned buildings named after a man or woman.

```python
women = pd.read_csv('/kaggle/input/women-representation-in-city-property-sanfrancisco/WomenRepresentaionInCityProperty-SanFrancisco.csv')
women.head()
```

```python
women.tail()
```

We note that the last columns will certainly not have a great impact in the analysis of this database. However, we will check them anyway.

```python
print(women.groupby(['Comments']).size())
```

```python
print(women.groupby(['Current Police Districts', 'Current Supervisor Districts', 'Analysis Neighborhoods', 'Neighborhoods',"SF Find Neighborhoods"]).size())
```

These results confirm that we can do without the last columns of this database. We will now analyse the others. 

```python
women_copy = women.drop(['Comments', 'Current Police Districts', 'Neighborhoods', 'Current Supervisor Districts', 'Analysis Neighborhoods','SF Find Neighborhoods'
], axis=1)
women_copy.head()
```

```python
sns.countplot(x ='Gender', data = women_copy)
print(women_copy.groupby(['Gender']).size())
```

First of all, we note that men were overwhelmingly represented. 
We are going to go into more detail, and see which sexes dominate by building category.

```python
print(women_copy.groupby(["Department/Source"]).size())
```

```python
sns.countplot(y = "Department/Source" ,hue ='Gender', data = women_copy)
```

We therefore notice that in this city women are poorly represented, with the exception of the port, where there seems to be equality between men and women.
