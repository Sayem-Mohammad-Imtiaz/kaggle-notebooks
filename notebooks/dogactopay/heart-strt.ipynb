```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../input/heart-disease-uci/heart.csv')
```

```python
df.columns
```

```python
X=df.drop('target',axis=1)
```

```python
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
sc=StandardScaler()
```

```python
scaled_df=pd.DataFrame(sc.fit_transform(X),columns=['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal',])
```

```python
scaled_df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_df, y, test_size=0.33, random_state=42)
```

```python
pc=PCA(n_components=2)
pc_df=pd.DataFrame(pc.fit_transform(X),columns=['feature1','feature2'])
pc_df=pd.concat([pc_df,y],axis=1)
```

```python
sns.scatterplot(x='feature1',y='feature2',data=pc_df,hue='target')
```

# KNeighbors

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.neighbors import KNeighborsClassifier
```

```python
acc=np.array([])
for i in range(1,190):
    model=KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train,y_train)
    pred=model.predict(X_test)
    acc=np.append(acc,accuracy_score(pred,y_test))

```

```python
np.argmax(acc)
```

```python
fig, ax = plt.subplots(figsize=(11.7, 8.27))
sns.lineplot(ax=ax,y=acc,x=range(1,190))
```

```python
model=KNeighborsClassifier(n_neighbors=148)
model.fit(X_train,y_train)
pred=model.predict(X_test)
print(classification_report(pred,y_test))
```

# SVM 

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
```

```python
model=SVC()
```

```python
model.fit(X_train,y_train)

pred=model.predict(X_test)

print(classification_report(pred,y_test))
```

```python
grid=GridSearchCV(estimator=SVC(),param_grid={'C': [0.1,1, 10, 100], 'gamma': [1,0.1,0.01,0.001], 'kernel': ('linear', 'rbf')},verbose=3,refit=True)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_estimator_
```

```python
grid_pred=grid.predict(X_test)
```

```python
print(classification_report(grid_pred,y_test))
```
