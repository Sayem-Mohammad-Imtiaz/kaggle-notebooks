# Basic Data Analysis

Lets find a list of all the songs and artists with rank 1

```python
rank1songs = []
rank1artists = []
for i in range(0, len(df)):
    if(df['rank'][i] == 1):
        rank1songs.append(df['song'][i])
        rank1artists.append(df['artist'][i])
```

```python
print(rank1artists)
```

```python
print(rank1songs)
```

Lets find out how many of The Beatles songs were rank 1

```python
print(df[df['artist'] == 'The Beatles']['song'].nunique())
print(df[df['artist'] == 'The Beatles']['song'].unique())
```
