#### Word Cloud in the Style of Indian Flag

```python
#OpenCV for the image reading
import cv2
#wordcloud lib 
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
text = str(list(data['State/UTs'])).replace(',', '').replace('[', '').replace("'", '').replace(']', '')
plt.rcParams['figure.figsize'] = (12, 12)
path=r"../input/indianflags/indian_sm.png"
im = cv2.imread(path)
im=cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
mask = np.array(im)
wordcloud_por = WordCloud(background_color="white", max_words=1000, mask=mask).generate(text)
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[8,6])
plt.imshow(wordcloud_por.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
```
