```python
data = pd.read_csv('/kaggle/input/covid19-in-india/covid_19_india.csv')
```

```python
data
```

```python
data.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational','Deaths','Confirmed'],axis = 1,inplace = True) 
```

```python
df = data.pivot_table('Cured',['Date'], 'State/UnionTerritory')
```

```python
df
```

```python
df.fillna(0, inplace=True)
```

```python
df.sort_values(list(df.columns),inplace=True)
```

```python
df=df.iloc[90:,]
```
