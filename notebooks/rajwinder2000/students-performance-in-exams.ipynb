# How group(race/ethnicity) affects the scores?

```python
df['total']=(df['math score']+df['reading score']+df['writing score'])/3
fig,axes=plt.subplots(1,4,figsize=(25,8))
sns.barplot(data=df,x='race/ethnicity',y='math score', ax = axes[0],ci=None)
axes[0].set_title("Maths Score")
plt.gcf().set_size_inches(12,6)
sns.barplot(data=df,x='race/ethnicity',y='writing score', ax = axes[1],ci=None)
axes[1].set_title("Writing Score")
plt.gcf().set_size_inches(12,6)
sns.barplot(data=df,x='race/ethnicity',y='reading score', ax = axes[2],ci=None)
axes[2].set_title("Reading Score")
plt.gcf().set_size_inches(12,6)
sns.barplot(data=df,x='race/ethnicity',y='total', ax = axes[3],ci=None)
axes[3].set_title("Total Score")
plt.gcf().set_size_inches(16,6)
```

The above observation indicates that, 'Group E' performed better in all subjects and 'Group A' performed worst

# Affect of lunch on the marks of the students

```python
df['total']=(df['math score']+df['writing score']+df['reading score'])/3
sns.set_style('whitegrid')
sns.barplot(data=df,x='lunch',y='total',ci=None)
plt.gcf().set_size_inches(10,6)
```
