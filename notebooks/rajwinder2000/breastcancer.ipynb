# MODELS(Predict whether the cancer is benign or malignant)

# Logistic Regression

```python
#'X' contain features which are used to train the model
X = df.iloc[:,1:]
X
```

```python
# 'Y' conatin target data,which is used to train the model
Y=Y= (df["diagnosis"]=="M").astype(np.int)
Y
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix,plot_confusion_matrix,accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.2, random_state=0)
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
clf=LogisticRegression()
clf.fit(X_train,Y_train)
pred=clf.predict(X_test)
logg_acc=accuracy_score(Y_test,pred)
# print(pred)
print("Accuracy of logistic model: ",logg_acc)
```

```python
print("REPORT: ")
print(confusion_matrix(Y_test,pred))
print(classification_report(Y_test,pred))
plot_confusion_matrix(clf,X_test,Y_test)
plt.grid(False)

```

# DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.2, random_state=0)
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)
clf = clf.fit(X_train,Y_train)
y_pred = clf.predict(X_test)
dtc_acc=accuracy_score(Y_test,y_pred)
print("REPORT:")
print("Accuracy of Decision_tree_Classifier: ",dtc_acc)
print(confusion_matrix(Y_test,y_pred))
print(classification_report(Y_test, y_pred))
plot_confusion_matrix(clf,X_test,Y_test)
plt.grid(False)
```

# KNN CLASSIFIER

```python
from sklearn.neighbors import KNeighborsClassifier

```

```python
# this helps in finding the value of 'n_neighbors' in KNN classifier
error_rate = []
  

for i in range(1, 40):
      
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(X_train, Y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != Y_test))
  
plt.figure(figsize =(10, 6))
plt.plot(range(1, 40), error_rate, color ='blue',
                linestyle ='dashed', marker ='o',
         markerfacecolor ='red', markersize = 10)
  
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
# I take value of 'n_neighbors'around 13,beacuse after this value error rate remain unchangeable 
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
clf=KNeighborsClassifier(n_neighbors = 13)
clf.fit(X_train,Y_train)
pred = clf.predict(X_test)
print("REPORT:")
knn_acc=accuracy_score(Y_test,pred)
print("Accuracy of knn model:",knn_acc)
print(confusion_matrix(Y_test, pred))
print(classification_report(Y_test, pred))
plot_confusion_matrix(clf,X_test,Y_test)
plt.grid(False)
```
