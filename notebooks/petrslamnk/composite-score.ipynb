```python
dataset = pd.read_csv("../input/marketing-data/marketing_data.csv")
```

```python
dataset["Monetary"] = dataset["MntWines"] + dataset["MntFruits"] + dataset["MntMeatProducts"] + dataset["MntFishProducts"] + dataset["MntSweetProducts"] + dataset["MntGoldProds"]
```

```python
dataset["Frequency"] = dataset["NumDealsPurchases"] + dataset["NumWebPurchases"] + dataset["NumCatalogPurchases"] + dataset["NumStorePurchases"] 
```

```python
dataset
```

```python
dataset_RFM = pd.DataFrame(dataset, columns=["ID", "Recency", "Frequency", "Monetary"])
```

```python
dataset_RFM["Recency"] = dataset_RFM["Recency"] * (-1)
```

```python
RFM = pd.DataFrame(dataset_RFM, ["Recency", "Frequency", "Monetary"])
```

```python
numerical_columns = RFM.drop("ID", axis = 1).columns
```

```python
dataset_RFM2 = dataset_RFM.copy()
```

```python
dataset_RFM2[numerical_columns] = scale.fit_transform(dataset_RFM2[numerical_columns])
dataset_RFM2.head()
```

```python
dataset_RFM2['Composite'] = dataset_RFM2.drop("ID", axis = 1).sum(axis=1)
dataset_RFM2.head()
```

```python
dataset_RFM2['Composite'] = max_scale.fit_transform(dataset_RFM2[['Composite']])
```

```python
dataset_RFM2
```

```python
# Create a letter score
dataset_RFM2['Letter Score'] = pd.qcut(dataset_RFM2['Composite'],6,labels=["F","E","D",'C','B','A'])
```
