```python
sns.set_style('whitegrid')
sns.set_context("paper")
sns.catplot(x = 'Year', data = N, kind = 'count')
plt.xticks(rotation = 90)
plt.show()
```

```python
sns.distplot(N["Runtime"], hist = True)
plt.show()
```

```python
sns.catplot(x = 'Age', data = N, kind = "count", order = ["7+", "13+", "16+", "18+", "all"])
plt.show()
```

```python
sns.relplot(x= "Runtime", y = "IMDb", data = N, kind = "scatter", marker = 'o', col = "Age", col_order = ["7+", "13+", "16+", "18+", "all"])
plt.show()
```

```python
sns.regplot(x = "IMDb", y = "Rotten Tomatoes", data = N)
plt.show()
```
