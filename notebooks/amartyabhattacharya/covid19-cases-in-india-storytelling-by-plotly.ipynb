# **Variation of Active Cases**

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['State/UTs'], y=df.Active, 
                        mode='lines+markers'))
```

```python
fig.update_layout(
    # Shows gray line without grid, styling fonts, linewidths and more
    xaxis=dict(
        showline=True,
        showgrid=False,
        showticklabels=True,
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        ticks='inside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='rgb(82, 82, 82)',
        ),
    ),
    # Turn off everything on y axis
    yaxis=dict(
        showgrid=False,
        zeroline=False,
        showline=True,
        showticklabels=True,
    ),
    autosize=True,
    margin=dict(
        autoexpand=False,
        l=100,
        r=20,
        t=110,
    ),
    showlegend=True,
    plot_bgcolor='Black'
)
```

###  Since the data is sorted with respect to the total cases, we can see that some of the tates with higher number of total cases has less number of active cases, so great job to those states!! 

### **Let us Visualize and compare the total cases vs active cases for each state and then we would understand what I told before.**

```python

fig = px.scatter(df, x="State/UTs", y="Total Cases", color="Active",)

fig.show()
```

### **We won't analyse the deaths in details since it's really sad event but yes we would like to know what has been the death ration in comparison to the number of deaths. Now death ratio is the major thing in concern and number of deaths don't give us a clear picture about the ratio. So Let us have a look at it.**

```python
fig = px.bar_polar(df, r="Death Ratio", theta="State/UTs", color="Deaths",
                   color_discrete_sequence= px.colors.sequential.Plasma_r,
                   title="Deaths vs Death Ratio Visualisation"
                  )
fig.show()
```

## *We can infer from this graph that the death ratio has been high in many places where the deaths recorded has been low.*

### **We do the same for the Active Ratio and total Active cases since, it will help us to visualize the proper scenario.**

```python
fig = px.bar_polar(df, r="Active Ratio", theta="State/UTs", color="Active",
                   color_discrete_sequence= px.colors.sequential.Plasma_r,
                   title="Active vs Active Ratio Visualization"
                  )
fig.show()
```

# Let us visualize and compare the active ratio for various states

```python
fig = px.bar(df, y='Active Ratio', x='State/UTs', color='Active')
# Put bar total value above bars with 2 values of precision
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
# Set fontsize and uniformtext_mode='hide' says to hide the text if it won't fit
fig.update_layout(uniformtext_minsize=8)
# Rotate labels 45 degrees
fig.update_layout(xaxis_tickangle=-45)
```

**IMPORTANT THING TO NOTE, ALTHOUGH MANY PLACES HAVE REALLY HIGH ACTIVE CASES THE ACTIVE RATIO IS REALLY LESS**

# Finally Let Us Visualize the overall picture of each states on the basis of Active cases, Discharge Ratio and Total Cases

```python
fig = px.scatter_3d(df, x='State/UTs', z='Active', y='Discharge Ratio',color = 'Total Cases',
                          opacity=0.5, width=900, height=800)
fig
```

### There is an assumption that the discharge ratio might be inversely proportional to the active cases, let us look at that 

```python
fig = px.scatter(df, x="State/UTs", y="Active", color="Discharge Ratio",)
fig
```

### **Let us note that our assumption is quite wrong because although in some states the discharge ratio is high, the number of cases is so high that it doesn't show the overall scenario. So we should consider the "Active Ratio" and look if it has any realtionship with the Discharge Ratio**

```python
fig = px.scatter(df, x="State/UTs", y="Active Ratio", color="Discharge Ratio",)
fig
```

### **YES! Our assumption was right ! For most of the cases, the increase in discharge ratio leads to the decrease in active ratio. Let us checkout if it's the same for death ratio too or not.** 

```python
fig = px.scatter(df, x="State/UTs", y="Death Ratio", color="Discharge Ratio",)
fig
```

### Sadly it didn't turn out to be exactly what we wanted, but yes it helps to realize the overall picture. 

# **And finally we will now checkout the active ratio and death ratio to realise the scene for all states**
