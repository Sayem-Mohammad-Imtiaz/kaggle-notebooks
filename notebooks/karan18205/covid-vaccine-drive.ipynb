```python
data = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv')
```

```python
countries = data['country'].unique()
print(countries)
print(len(countries))
```

```python
data['total_vaccinations'] = data['total_vaccinations'].fillna(0)
data['daily_vaccinations_per_million'] = data['daily_vaccinations_per_million'].fillna(0)
data.head()
```

```python
totalVacc = sum(data['total_vaccinations'])
totalVacc
```

```python
# Total Vaccinations for countries

total = {}
for country in countries:
    sumC = 0
    for index, row in data.iterrows():
        if row['country'] == country:
            sumC = sumC + row['total_vaccinations']
    total[country] = sumC
total
```

```python
# Countries using which vaccines
data['total_vaccinations'] = data['total_vaccinations'].fillna(0)
data_VaccineType = data[['country','vaccines']]
data_VaccineType_dict = {}
for country in countries:
    vaccinesPcountry = []
    for index, row in data_VaccineType.iterrows():
        if row['country'] == country:
            if row['vaccines'] not in vaccinesPcountry:
                vaccinesPcountry.append(row['vaccines'])
    data_VaccineType_dict[country] = vaccinesPcountry
data_VaccineType_dict
```

```python
# Vaccination Progression
import matplotlib.pyplot as plt

data_VaccineProgression = data[['country','daily_vaccinations_per_million']]
data_VaccineProgression_dict = {}
for country in countries:
    vaccinesProgressioncountry = []
    for index, row in data_VaccineProgression.iterrows():
        if row['country'] == country:
            vaccinesProgressioncountry.append(row['daily_vaccinations_per_million'])
    data_VaccineProgression_dict[country] = vaccinesProgressioncountry
```

```python
for country in countries:
    plt.plot(data_VaccineProgression_dict[country])
    plt.ylabel('Vaccines '+country)
    plt.xlabel('Days')
    plt.show()
```
