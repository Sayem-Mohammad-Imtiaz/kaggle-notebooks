**Silicon Valley Diversity Data**

**Overview**
         Silicon Valley is known for its immense diversity in it's workforce. A lot of companies are making sure that they employ skilled employee's of different ethinicity and gender, thus providing equal opportunities. The data source used for this analysis is accounted for few of the top employers in the Silicon Valley for the year 2016.
         
 The scope of this analysis is to look into few of the diversity measures employed by a lot of companies.
 
 i.To have the overall male to female workforce ratio to an equal 50:50 in the company.
 
 ii. To have the male to female workforce ratio to an equal 50:50 in each department of the company.
 
 iii. Atleast 40% of the workforce composes of ethnically diverse employees.
            
**Analysis**
Tableau Dashboard Link depicting the Analysis:  https://public.tableau.com/profile/chinmayi8680#!/vizhome/Kaggle-DiversityinSiliconValley/DiverstiyinSiliconValleyCompanies

i. Keeping an overall male to female workforce ratio to an equal 50:50.
     Gender diversity implies equal employment opportunity for both male and female skilled workers. 
     
 Tableau worksheet link for the above analysis :  https://public.tableau.com/profile/chinmayi8680#!/vizhome/Kaggle-DiversityinSiliconValley/Case1

23andMe company seems to be doing the best they can to achieve gender diversity in their employment. Their male to female employee ratio is almost 50:50. The company is closely followed by Pinterest, Intuit  and Airbnb in achieving this ratio.  View and Nvidia are the two companies who really need to restructure their gender diversity employment policies. Both the companys have female workforce of less that 18%, with men taking away most of the jobs.
     
ii. A lot of departments have a specific gender dominant workforce. Being inclusive of other skilled workers from different gender should be encouraged by a lot of Tech companies.

Tableau worksheet link for the above analysis:  https://public.tableau.com/profile/chinmayi8680#!/vizhome/Kaggle-DiversityinSiliconValley/Case2

The percentage shown against each bar in the viz, is the percentage difference between the male and female workforce count in that particular department of the company. Lesser percentage difference implies higher gender equality rate. 

In the viz, you can drill down to a specific company, by using the company filter. Or you can update your viz to any specific percentage difference window values. 

iii. Atleast 40% of the company's workforce should be ethnically diverse. Even though this is not an ideal case, as the 40% has few major sub categories like Asian, African, Pacific, Indian, Latino etc, this can be accounted as a stepping stone towards a more structured diverse employment of skilled workers.

Most of the companies seem to be inclusive of employing skilled workers of different race. Two companies that have the least ethnic diversity are HPE and HP Inc., with less that 30% of their overall workforce being non-white.
