```python
data=pd.read_csv('/kaggle/input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')
```

```python
data.head()
```

```python
#change string value to integer value ex: male=0 and female=1
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
labelencoder=LabelEncoder()
data['gender']=labelencoder.fit_transform(data['gender'])
data['ever_married']=labelencoder.fit_transform(data['ever_married'])
data['work_type']=labelencoder.fit_transform(data['work_type'])
data['Residence_type']=labelencoder.fit_transform(data['Residence_type'])
data['smoking_status']=labelencoder.fit_transform(data['smoking_status'])
#onehotencoder=OneHotEncoder(cathegorical_features=[2])
```

```python
data.head()
```

```python
#filling value for nan in bmi column its mean value
mean_value=data['bmi'].mean()
data['bmi'].fillna(value=mean_value, inplace=True)
```

```python
data.head()
```

```python
X=data[['gender', 'age', 'hypertension', 'heart_disease', 'ever_married', 'work_type', 'Residence_type', 'avg_glucose_level', 'bmi', 'smoking_status']]
y=data['stroke']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(X,y)
```

```python
y_pred=tree.predict(X_test)
```

```python
#100 accuracy i achived
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)*100
```
