we will start now the process of prediciton.
We will define metrics used for prediction.
And we will define the label column to predict

```python
x=df.loc[:, df.columns != 'price_range']

y=df['price_range']
```

We will use now the decision tree regressor method for classification

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=30)

############## Regression
from sklearn.tree import DecisionTreeRegressor
dt_model = DecisionTreeRegressor(random_state=1)

# Fit the model
dt_model.fit(x_train,y_train)
from sklearn.metrics import accuracy_score
y_pred=dt_model.predict(x_test)
print('Accuracy=',accuracy_score(y_pred,y_test))
```

now we use KNN neighbors algorithm for prediction

```python
from sklearn.neighbors import KNeighborsClassifier
n_neighbors=30   
knn=KNeighborsClassifier(n_neighbors)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print('Accuracy=',accuracy_score(y_pred,y_test))
```

We classify data now using the naive bayes method

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
y_pred = gnb.fit(x_train, y_train).predict(x_test)
print('Accuracy=',accuracy_score(y_pred,y_test))
```

As see above, using the KNN neighbors, we get the best accuracy value when predicting the range of price of mobile phones
