```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import xgboost

classifiers = {
    "LogisiticRegression": LogisticRegression(),
    "KNearest": KNeighborsClassifier(),
#     "Support Vector Classifier": SVC(),
#     "DecisionTreeClassifier": DecisionTreeClassifier(),
    "RandomForestClassifier":RandomForestClassifier(),
    'XgBoost':xgboost.XGBClassifier()
}
```

```python
from sklearn.model_selection import cross_val_score


for key, classifier in classifiers.items():
    classifier.fit(x_train, y_train)
    training_score = cross_val_score(classifier, x_train, y_train, cv=2)
    print("Classifiers: ", classifier.__class__.__name__, "Has a training score of", round(training_score.mean(), 2) * 100, "% accuracy score")
```

```python
print("starting grid serach cv on logistic regression....")
from sklearn.model_selection import GridSearchCV


# Logistic Regression 
log_reg_params = {"penalty": ['l1', 'l2'], 'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}


grid_log_reg = GridSearchCV(LogisticRegression(), log_reg_params,cv=2)
grid_log_reg.fit(x_train, y_train)
#We automatically get the logistic regression with the best parameters.
log_reg = grid_log_reg.best_estimator_

print("starting grid serach cv on k nearest neibhours...")

knears_params = {"n_neighbors": list(range(4,7,1)), 'algorithm': ['auto', 'ball_tree', 'kd_tree']}

grid_knears = GridSearchCV(KNeighborsClassifier(), knears_params,cv=2)
grid_knears.fit(x_train, y_train)
# KNears best estimator
knears_neighbors = grid_knears.best_estimator_
```

```python
print("starting grid serach cv on random forest...")

rfc=RandomForestClassifier(random_state=42)

param_grid = { 
    'n_estimators': [100, 200],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5],
    'criterion' :['gini', 'entropy']
}
CV_rfc = GridSearchCV(estimator=rfc, param_grid=param_grid,cv=2)
CV_rfc.fit(x_train, y_train)
random_tree_clf=CV_rfc.best_estimator_
```

```python
# n_estimators = [int(x) for x in np.linspace(start = 10, stop = 200, num = 10)]
# max_features = ['auto', 'sqrt']
# max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
# max_depth.append(None)
# min_samples_split = [2, 5, 10]
# min_samples_leaf = [1, 2, 4]
# bootstrap = [True, False]
xgb_grid = {'n_estimators': [200],
               'max_features': ['sqrt'],
               'max_depth': [5],
               'min_samples_split': [10],
               'min_samples_leaf': [4],
               'bootstrap': [True]}
xgb=xgboost.XGBClassifier()
grid_xgb = GridSearchCV(estimator=xgb, param_grid=xgb_grid,cv=2)
grid_xgb.fit(x_train, y_train)
xgb_classifier=grid_xgb.best_estimator_
```

```python
print("logestic regression best parameter",grid_log_reg.best_params_)
print("knears neibhours best parameter",grid_knears.best_params_)
print("Random Forest Classifier  best parameter",CV_rfc.best_params_)
print("xgboost classifier best parameter",grid_xgb.best_params_)
```

```python
log_reg_score = cross_val_score(log_reg, x_train, y_train, cv=2)
print('Logistic Regression Cross Validation Score: ', round(log_reg_score.mean() * 100, 2).astype(str) + '%')


knears_score = cross_val_score(knears_neighbors, x_train, y_train, cv=2)
print('Knears Neighbors Cross Validation Score', round(knears_score.mean() * 100, 2).astype(str) + '%')


random_forest_score = cross_val_score(random_tree_clf, x_train, y_train, cv=2)
print('Random forest Classifier Cross Validation Score', round(random_forest_score.mean() * 100, 2).astype(str) + '%')

xgb_score = cross_val_score(xgb_classifier, x_train, y_train, cv=2)
print('DecisionTree Classifier Cross Validation Score', round(xgb_score.mean() * 100, 2).astype(str) + '%')
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_predict

log_reg_pred = cross_val_predict(log_reg, x_train, y_train, cv=5,
                             method="decision_function")

knears_pred = cross_val_predict(knears_neighbors, x_train, y_train, cv=5)


random_forest_pred = cross_val_predict(random_tree_clf, x_train, y_train, cv=5)
xgb_pred = cross_val_predict(xgb_classifier, x_train, y_train, cv=5)
```

```python
log_fpr, log_tpr, log_thresold = roc_curve(y_train, log_reg_pred)
knear_fpr, knear_tpr, knear_threshold = roc_curve(y_train, knears_pred)
random_forest_fpr, random_forest_tpr, random_forest_threshold = roc_curve(y_train, random_forest_pred)
xgb_fpr,xgb_tpr,xgb_threshold=roc_curve(y_train,xgb_pred)
def graph_roc_curve_multiple(log_fpr, log_tpr, knear_fpr, knear_tpr,random_forest_fpr,random_forest_tpr,xgb_fpr,xgb_tpr):
    plt.figure(figsize=(16,8))
    plt.title('ROC Curve \n Top 4 Classifiers', fontsize=18)
    plt.plot(log_fpr, log_tpr, label='Logistic Regression Classifier Score: {:.4f}'.format(roc_auc_score(y_train, log_reg_pred)))
    plt.plot(knear_fpr, knear_tpr, label='KNears Neighbors Classifier Score: {:.4f}'.format(roc_auc_score(y_train, knears_pred)))
    plt.plot(random_forest_fpr, random_forest_tpr, label='Random forest Classifier Score: {:.4f}'.format(roc_auc_score(y_train, random_forest_pred)))
    plt.plot(xgb_fpr,xgb_tpr,label="Xgb Classifier Score: {:.4f}".format(roc_auc_score(y_train,xgb_pred)))
    plt.plot([0, 1], [0, 1], 'k--')
    plt.axis([-0.01, 1, 0, 1])
    plt.xlabel('False Positive Rate', fontsize=16)
    plt.ylabel('True Positive Rate', fontsize=16)
    plt.annotate('Minimum ROC Score of 50% \n (This is the minimum score to get)', xy=(0.5, 0.5), xytext=(0.6, 0.3),
                arrowprops=dict(facecolor='#6E726D', shrink=0.05),
                )
    plt.legend()
    
graph_roc_curve_multiple(log_fpr, log_tpr, knear_fpr, knear_tpr, random_forest_fpr,random_forest_tpr,xgb_fpr,xgb_tpr)
plt.show()
```
