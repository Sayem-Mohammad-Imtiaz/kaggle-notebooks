We are using Pandas framework for extract our whole range of data. I gaved "df"  name for data set that include world happines report.And I am using "df.columns" method for seeing what are the columns, you can say also features, of my data set.

```python
df = pd.read_csv('../input/2017.csv')
df.columns
```

Here, I would like to change some of the column names. Because some of them contain gaps and a couple dots,  this can be problem when I want to use significant methods. 
E.g ( df.Economy..GDP.per.Capita.plot() ) this cod line giving an sytax error. To solve this, we change column names.


```python
df.columns =[" ".join(each.split("..")) if (len(each.split("..")) > 1) else each for each in df.columns ]
df.columns =["_".join(each.split(".")) if (len(each.split(".")) > 1) else each for each in df.columns ]
df.columns =["_".join(each.split(" ")) if (len(each.split(" ")) > 1) else each for each in df.columns ]

df.rename(columns={'Economy_GDP_per_Capita_': 'Economy_GDP_per_Capita'}, inplace=True)
df.rename(columns={'Trust_Government_Corruption_': 'Trust_Government_Corruption'}, inplace=True)
df.rename(columns={'Health_Life_Expectancy_': 'Health_Life_Expectancy'}, inplace=True)
```

"corr" method, lists for us the correlation between features. Correlation is a statistical technique that can show whether and how strongly pairs of features(columns) are related to each other. The relationship isn't perfect. Houses of the same area (square metre) vary in price, and you can easily think of two houses you know where the narrower one is more expensive than the larger one.  Because it does not consist of only two features. Correlation can tell you just how much of the variation in houses'' price is related to their quantitative of living area.

Although this correlation is fairly obvious your data may contain unsuspected correlations. You may also suspect there are correlations, but don't know which are the strongest. An intelligent correlation analysis can lead to a greater understanding of your data.

~(-1) represents: Negative correlation/	Direct proportion 

0 represents: No relations

~(+1) represents: Positive correlation/	Inverse proportion 

```python
df.corr()
```

I am creating new data frames from our default data set and I am giving new frame names for each country.

```python
norway = df[df.Country == "Norway"]                      #Rank 1
denmark = df[df.Country == "Denmark"]                    #Rank 2
iceland = df[df.Country == "Iceland"]                    #Rank 3
finland = df[df.Country == "Finland"]                    #Rank 5
sweden = df[df.Country == "Sweden"]                      #Rank 9

turkey = df[df.Country == "Turkey"]                      #Rank 69
syria = df[df.Country == "Syria"]                        #Rank 117
israel = df[df.Country == "Israel"]                      #Rank 11
qatar = df[df.Country == "Qatar"]                        #Rank 35
egypt = df[df.Country == "Egypt"]                        #Rank 104
```

After I set up the figure size, I would like to observe dependency level between happiness rank and the trust government corruption. It is clear that Syria has more trust rate compared to other, but it still has the worst happiness rank. However, Iceland has the lowest trust rate, but it ranks third.

```python
plt.figure(figsize=(20,10))

plt.bar(norway.Happiness_Rank, norway.Trust_Government_Corruption, color = 'navy', label = 'Norway')
plt.bar(denmark.Happiness_Rank, denmark.Trust_Government_Corruption, color = 'red', label = 'Denmark')
plt.bar(iceland.Happiness_Rank, iceland.Trust_Government_Corruption, color = 'grey', label = 'Iceland')
plt.bar(finland.Happiness_Rank, finland.Trust_Government_Corruption, color = 'lightblue', label = 'Finland')
plt.bar(sweden.Happiness_Rank, sweden.Trust_Government_Corruption, color = 'yellow', label = 'Sweden')

plt.bar(turkey.Happiness_Rank, norway.Trust_Government_Corruption, label = 'Turkey')
plt.bar(syria.Happiness_Rank, denmark.Trust_Government_Corruption, label = 'Syria')
plt.bar(israel.Happiness_Rank, iceland.Trust_Government_Corruption, label = 'Israel')
plt.bar(qatar.Happiness_Rank, finland.Trust_Government_Corruption, label = 'Qatar')
plt.bar(egypt.Happiness_Rank, sweden.Trust_Government_Corruption, label = 'Egypt')

plt.legend(loc = 'upper right')
plt.title('Effect of Trust Government against Happiness Rank')
plt.xlabel('Happiness Rank')
plt.ylabel('Trust Government Corruption')

plt.show()
```

Using the same figure size, I visualized the relation between freedom and economic GDP. And we can easily observe, there is a direct proportion between economic situation and freedom. We can understand that money provides freedom. When you have much money, you are more free.

```python
plt.figure(figsize=(20,10))

plt.scatter(norway.Economy_GDP_per_Capita, norway.Freedom, color = 'navy', label = 'Norway')
plt.scatter(denmark.Economy_GDP_per_Capita, denmark.Freedom, color = 'red', label = 'Denmark')
plt.scatter(iceland.Economy_GDP_per_Capita, iceland.Freedom, color = 'grey', label = 'Iceland')
plt.scatter(finland.Economy_GDP_per_Capita, finland.Freedom, color = 'lightblue', label = 'Finland')
plt.scatter(sweden.Economy_GDP_per_Capita, sweden.Freedom, color = 'yellow', label = 'Sweden')

plt.scatter(turkey.Economy_GDP_per_Capita, turkey.Freedom, color = 'firebrick', label = 'Turkey')
plt.scatter(syria.Economy_GDP_per_Capita, syria.Freedom, color = 'magenta', label = 'Syria')
plt.scatter(israel.Economy_GDP_per_Capita, israel.Freedom, color = 'slateblue', label = 'Israel')
plt.scatter(qatar.Economy_GDP_per_Capita, qatar.Freedom, color = 'coral', label = 'Qatar')
plt.scatter(egypt.Economy_GDP_per_Capita, egypt.Freedom, color = 'olive', label = 'Egypt')

plt.legend(loc = 'upper left')
plt.title('Effect of GDP against Freedom')
plt.xlabel('GDP')
plt.ylabel('Freedom')

plt.show()
```
