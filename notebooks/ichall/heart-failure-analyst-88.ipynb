```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv')
df.head()
```

```python
df.info()
```

```python
sns.countplot(hue=df['DEATH_EVENT'],x=df['anaemia'])
```

```python
sns.countplot(df['high_blood_pressure'],hue=df['DEATH_EVENT'])
```

```python
X = df.drop(columns='DEATH_EVENT')
y = df['DEATH_EVENT']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
dtr = DecisionTreeClassifier()
rfc = RandomForestClassifier()
```

```python
dtr.fit(X_train,y_train)
rfc.fit(X_train,y_train)
```

```python
y_dtr = dtr.predict(X_test)
y_rfc = rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

print(accuracy_score(y_test,y_dtr))
print('\n')
print(classification_report(y_test,y_dtr))
print('-'*80)
print(accuracy_score(y_test,y_rfc))
print('\n')
print(classification_report(y_test,y_rfc))
print('-'*80)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators':[200,300],'criterion':['entropy','gini'],'max_depth':[1,2,3,4,5]}
grid = GridSearchCV(RandomForestClassifier(),param_grid,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid = grid.best_estimator_
```

```python
grid.fit(X_train,y_train)
```

```python
y_grid=grid.predict(X_test)
```

```python
print(accuracy_score(y_test,y_grid))
print('\n')
print(classification_report(y_test,y_grid))
```
