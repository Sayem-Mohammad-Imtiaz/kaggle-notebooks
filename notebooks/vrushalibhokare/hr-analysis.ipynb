```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df= pd.read_csv('/kaggle/input/hr-analytics/HR_comma_sep.csv')
df.head()
```

```python
df.groupby('left').mean()
```

- **Satisfaction Level**: Satisfaction level seems to be relatively low (0.44) in employees leaving the firm vs the retained ones (0.66)
- **Average Monthly Hours**: Average monthly hours are higher in employees leaving the firm (199 vs 207)
- **Promotion Last 5 Years**: Employees who are given promotion are likely to be retained at firm

```python
pd.crosstab(df.salary,df.left).plot(kind='bar')
```

In above graph, the employee with higher salary are likely to not leave the firm

```python
pd.crosstab(df.Department,df.left).plot(kind='bar')
```

By doing analysis, we choose following variables: 
satisfaction_level, promotion_last_5years, average_montly_hours, salary

```python
df1=df[['satisfaction_level','promotion_last_5years','average_montly_hours','salary']]
df1.head()
```

```python
sal_dummies= pd.get_dummies(df1['salary'],prefix="salary")
df_with_dummies = pd.concat([df1,sal_dummies],axis='columns')
```

```python
df_with_dummies.drop('salary',axis=1, inplace=True)
```

```python
df_with_dummies.head()
```

```python
X=df_with_dummies
y=df['left']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=46)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred= model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={"C":np.linspace(.01, 100, 10)}
model=LogisticRegression(max_iter=15000)
```

```python
logreg_cv=GridSearchCV(model,params,cv=10,scoring= 'accuracy')
logreg_cv.fit(X_train,y_train)
```

```python
logreg_cv.best_params_
```

```python
lr=LogisticRegression(C=0.01, penalty= 'l2')
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```
