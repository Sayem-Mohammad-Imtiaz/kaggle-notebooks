**MATPLOTLIB**

Matplot, verileri grafiklere dökmek için kullandığımız bir python kütüphanesidir. En kolay ve temel grafikler line, scatter ve histogramlardır.

* Eksenlerimizden birisi zaman ise line, diğer grafiklere nazaran daha iyi sonuçlar verecektir.
* İki değişken arasındaki korelasyonu görmek istiyorsak scatter'ı kullanabiliriz.
* Sayısal bir verinin dağılımını görmek istiyorsak histogram'ı kullanabiliriz.
* Özelleştirme : Grafiklerin renkleri, çizgi kalınlıkları, opaklığı gibi özelliklerini ayarlayabiliriz.

# 2.Python Data Science Toolbox

**User Defined Function**

```python
def tuble_ex():
    """yorum satırı"""
    t = (1,2,3)
    return t 
print(tuble_ex())
```

**Scope**

```python
x = data.Happiness_Score
def f():
    y = 2 * x
    return y
print(f())

```

**Nested Function**

```python
def cember_cevre(r):
    def carp(pi=3.14):
        sonuc = 2 * pi
        return sonuc 
    return carp()*r
print(cember_cevre(2))
    
    
```

**Default and Flexible Arguments**

```python
def f(*args):
    for i in args:
          print(i)
f(10)  

def k(**kwargs):
    for key, value in kwargs.items():
        print(key, " ", value)
k(country = 'spain', capital = 'madrid', populaiton = 123456)      

        
```

```python
number_list = [1,2,3]
y = map(lambda x:x**2,number_list)
print(list(y))
```

```python
list1 = [1,2,3,4]
list2 = [5,6,7,8]
z = zip(list1,list2)
print(z)
z_list = list(z)
print(z_list)
```

**List Comprehension**

```python
num1 = [1,2,3]
num2 = [i+1 for i in num1]
print(num2)
```

```python
num1 = [5,10,15]
num2 = [i**2 if i == 10 else i - 5 if i < 7 else i + 5 for i in num1]
print(num2)
```
