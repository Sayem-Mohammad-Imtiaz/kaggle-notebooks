```python
sns.catplot(x="race/ethnicity",y="Total_Marks",kind="box",hue="lunch",showmeans=True,data=NDf)
#As it can be seen from the following that for all the races the median and almost all statistics are below for "free/reduced" are low.
# It is also evident that Group E have the hightst median as compared to all other Gropus.
#It shows a direct relation between the lunch and total marks.
```

```python
sns.catplot(y="Total_Marks",hue="gender",x="race/ethnicity",kind="box",data=NDf)
#As it can be seen that for each clasee median value for the female is higher in each class.
#thus it can be concluded that an Avg girl in each class have more marks as compared to avg boy. 
```
