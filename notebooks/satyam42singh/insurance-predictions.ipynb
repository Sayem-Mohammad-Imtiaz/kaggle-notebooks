### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
X=df.drop("Response",axis=1)
y=df["Response"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
y_train
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
X_train.head()
```

```python
from imblearn.combine import SMOTEENN
smk = SMOTEENN()
X_res,y_res=smk.fit_resample(X_train,y_train)
X_res.shape,y_res.shape
```

```python
model=LogisticRegression()
model.fit(X_res,y_res)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
y_predict=model.predict_proba(X_test)
```

```python
y_predict
```

```python
from sklearn import metrics
cm=metrics.confusion_matrix(y_test, y_pred, labels=[1, 0])

df_cm = pd.DataFrame(cm, index = [i for i in ["1","0"]],
                  columns = [i for i in ["Predict 1","Predict 0"]])
plt.figure(figsize = (7,5))
sns.heatmap(df_cm, annot=True,fmt='g')
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test,y_predict[:,1])
```

```python
fpr,tpr,threshold=metrics.roc_curve(y_test,y_predict[:,1])
plt.plot(fpr,tpr)
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.0])
plt.title("ROC curve for Logisticclassifier")
plt.xlabel("FPR(1-specificity)")
plt.ylabel("TPR(Senstivity)")
plt.grid(True)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfcl=RandomForestClassifier(n_estimators=15,random_state=42,max_features=11)
rfcl=rfcl.fit(X_res,y_res)
```

```python
y_rf=rfcl.predict(X_test)
```

```python
print(rfcl.score(X_test,y_test))
```

```python
y_rf=rfcl.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.classification_report(y_test, y_rf))
```

```python
cm=metrics.confusion_matrix(y_test,y_rf,labels=[0,1])
df_cm=pd.DataFrame(cm,index=[i for i in ["No","Yes"]],
                  columns=[i for i in ["No","Yes"]])
plt.figure(figsize=(7,5))
sns.heatmap(df_cm,annot=True,fmt="g")
```

```python
y_ptpr=rfcl.predict_proba(X_test)[:,1]
y_ptpr
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import auc

plt.style.use("ggplot")

# Compute fpr, tpr, thresholds and roc auc
fpr, tpr, thresholds = roc_curve(y_test, y_ptpr)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % roc_auc)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate or (1 - Specifity)')
plt.ylabel('True Positive Rate or (Sensitivity)')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.grid(True)
```
