```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

##**Fatal Police shooting**##

```python

data_set = pd.read_csv("../input/police-shootings/database.csv")
data_set
```

**Total count per races**

```python
data_set['race'].value_counts()
```

```python
wAttack = len(data_set.loc[(data_set['race'] == 'W') & (data_set['threat_level'] == 'attack')])
bAttack = len(data_set.loc[(data_set['race'] == 'B') & (data_set['threat_level'] == 'attack')])

wPercent = wAttack / 1041 *100
bPercent = bAttack / 542 *100

print("W with attack threats: %d, %0.1f%%" % (wAttack, wPercent))
print("B with attack threats: %d, %0.1f%%" % (bAttack, bPercent))
```

**The frequency count of threat levels**

```python
data_set['threat_level'].value_counts()
```

**We want to focus on the "B" race without an "attack" level of threat**

```python
data_set.groupby(['race', 'threat_level']).size().reset_index(name="Count")
```

```python
BUnarmed = len(data_set.loc[(data_set['race'] == 'B') & (data_set['armed'] == 'unarmed') & (data_set['threat_level'] != 'attack')])
WUnarmed = len(data_set.loc[(data_set['race'] == 'W') & (data_set['armed'] == 'unarmed') & (data_set['threat_level'] != 'attack')])

wUnArmedPerc = WUnarmed / 1041 * 100
bUnArmedPerc = BUnarmed / 542 * 100

print('Black men who were killed unarmed: %d, %0.1f%%' % (BUnarmed, bUnArmedPerc))
print('White men who were killed unarmed: %d, %0.1f%%' % (WUnarmed, wUnArmedPerc))

```
