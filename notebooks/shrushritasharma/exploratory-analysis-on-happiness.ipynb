```python
# Reading the dataset
data_2015 = pd.read_csv('../input/world-happiness/2015.csv')
data_2016 = pd.read_csv('../input/world-happiness/2016.csv')
data_2017 = pd.read_csv('../input/world-happiness/2017.csv')
data_2018 = pd.read_csv('../input/world-happiness/2018.csv')
data_2019 = pd.read_csv('../input/world-happiness/2019.csv')
```

```python
# Merging the datasets
data_2015 = data_2015.rename(columns={
'Country':'Country_2015','Region':'Region_2015','Happiness Rank':'Rank_2015','Happiness Score':'Score_2015',
'Standard Error':'Error_2015','Economy (GDP per Capita)':'GDP_2015','Family':'Family_2015','Health (Life Expectancy)':'Health_2015',
'Freedom':'Freedom_2015','Trust (Government Corruption)':'Govt_2015','Generosity':'Generosity_2015','Dystopia Residual':'Dystopia_2015'})

data_2016=data_2016.rename(columns={
'Country':'Country_2016','Region':'Region_2016','Happiness Rank':'Rank_2016','Happiness Score':'Score_2016',
'Lower Confidence Interval':'LConf_2016','Upper Confidence Interval':'UConf_2016','Economy (GDP per Capita)':'GDP_2016',
'Family':'Family_2016','Health (Life Expectancy)':'Health_2016','Freedom':'Freedom_2016',
'Trust (Government Corruption)':'Govt_2016','Generosity':'Generosity_2016','Dystopia Residual':'Dystopia_2016'})

data_2017=data_2017.rename(columns={
'Country':'Country_2017','Happiness.Rank':'Rank_2017','Happiness.Score':'Score_2017','Whisker.high':'UConf_2017','Whisker.low':'LConf_2017',
'Economy..GDP.per.Capita.':'GDP_2017','Family':'Family_2017','Health..Life.Expectancy.':'Health_2017','Freedom':'Freedom_2017',
'Generosity':'Generosity_2017','Trust..Government.Corruption.':'Govt_2017','Dystopia.Residual':'Dystopia_2017'})

data_2018=data_2018.rename(columns={
'Overall rank':'Rank_2018','Country or region':'Country_2018','Score':'Score_2018','GDP per capita':'GDP_2018','Social support':'Family_2018',
'Healthy life expectancy':'Health_2018','Freedom to make life choices':'Freedom_2018','Generosity':'Generosity_2018','Perceptions of corruption':'Govt_2018'})

data_2019=data_2019.rename(columns={
'Overall rank':'Rank_2019','Country or region':'Country_2019','Score':'Score_2019','GDP per capita':'GDP_2019','Social support':'Family_2019',
'Healthy life expectancy':'Health_2019','Freedom to make life choices':'Freedom_2019','Generosity':'Generosity_2019','Perceptions of corruption':'Trust_2019'})

data = pd.concat([data_2015,data_2016,data_2017, data_2018, data_2019], axis=1)
#data.info()
```
