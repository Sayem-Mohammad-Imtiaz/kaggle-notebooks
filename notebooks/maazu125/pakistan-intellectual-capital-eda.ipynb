```python
dataset=pd.read_csv("../input/pakistanintellectualcapitalcs/Pakistan Intellectual Capital - Computer Science - Ver 1.csv", encoding = "ISO-8859-1")
dataset.head()
```

```python
# Universities Ranking by Teachers
import matplotlib.pyplot as plt
university="University Currently Teaching"
plt.rcParams['figure.figsize']=(12,50)
dataset[university].value_counts().sort_values().plot(kind="barh")
plt.show()
```

```python
#Computer science/IT professors from 89 different universities of Pakistan
dataset.Department.groupby(dataset.Department).count().plot(kind="bar")
```

```python
#Which area of interest/expertise is in abundance in Pakistan and where we need more people?
Specialization = "Area of Specialization/Research Interests"
dataset[Specialization].value_counts().sort_values()[::-1][:20].plot(kind="bar")
```

```python
#Which province have most professors?
Province = "Province University Located"
dataset[Province].value_counts().sort_values().plot(kind="bar")


```

```python
#Which country and university hosted majority of our teachers?
Country = "Country"
dataset[Country].value_counts().sort_values().plot(kind="bar")
```

```python
#Terminal degree of Teachers
Terminal_Degree = "Terminal Degree"
dataset[Terminal_Degree].value_counts().sort_values().plot(kind="bar")
```

```python
#Designation of Teachers
Designation = "Designation"
dataset[Designation].value_counts().sort_values().plot(kind="bar")
```
