```python
df_object = data.select_dtypes(include="O")
df_number = data.select_dtypes(exclude="O")

df_object
```

```python
label = LabelEncoder()
df_object = df_object.astype(str).apply(label.fit_transform)

data = pd.concat([df_object, df_number], axis = 1)
data
```

```python
X = data.drop(['stroke'],axis = 1)
y = data.stroke
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(X.corr(), linecolor='black',cmap='YlGnBu', linewidth=1, annot=True)
```

```python
scaler = StandardScaler(copy=True, with_mean=True, with_std=True)
X = scaler.fit_transform(X)
X 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=44, shuffle =True)

#Splitted Data
print('X_train shape is ' , X_train.shape)
print('X_test shape is ' , X_test.shape)
print('y_train shape is ' , y_train.shape)
print('y_test shape is ' , y_test.shape)
```

```python
LClassifier = LazyClassifier(verbose=0,ignore_warnings=True, custom_metric=None)
model,predictions = LClassifier.fit(X_train, X_test, y_train, y_test)
```
