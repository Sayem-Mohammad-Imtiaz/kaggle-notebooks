# 3. Groupby (for categorical variables)

### Use groupby method for columns data affect mostly in scores of students to analyze properly

```python
data=df_1.groupby(['gender','lunch','test preparation course'], as_index=False).mean() 
data
```

From above table,
* we can say that when we       consumed  more hygienic       and good quality of           food(lunch) as well as         after taking                   preparation course in         that subject we can get       a better results.     

* for math subject, males       seems to be more sharp and     can perform better than       females. 

* However, when it comes to     observe deeply and             performance (reading and       writing) female likely can     do in a more well manner       than boys or male. 

****As of now, we conclude that the features that can affect mostly to the performance of the students in exam is as below;
1. Gender
2. Lunch
3. Test preparation course
