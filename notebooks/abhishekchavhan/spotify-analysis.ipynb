```python
# Import the basic tools : 
import pandas as  pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import the data in encoded format : 
filename='/kaggle/input/top50spotify2019/top50.csv'
data=pd.read_csv(filename,encoding='ISO-8859-1')
data.drop('Unnamed: 0',inplace=True,axis=1)
```

```python
## Lets predict the popularity now:

data['Popularity'].plot();
```

```python
data.dtypes
```

```python
# Lets convert the object type into categorical data type :
for label,content in data.items():
    if pd.api.types.is_string_dtype(content):
        print(label)
```

```python
for label,content in data.items():
    if pd.api.types.is_string_dtype(content):
        data[label] = content.astype('category').cat.as_ordered()
```

```python
data.dtypes
```

```python
for label,content in data.items():
    if pd.api.types.is_categorical_dtype(content):
        data[label] = pd.Categorical(content).codes + 1
```

```python
X = data.drop('Popularity',axis=1)
y = data['Popularity']
```

```python
# Lets import modelling tools:
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.model_selection import train_test_split,RandomizedSearchCV,GridSearchCV
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
model = GradientBoostingRegressor()
model.fit(X_train,y_train)
```

```python
from xgboost import XGBRegressor
```

```python
model = XGBRegressor()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
predictions = model.predict(X_test)
mae_1 = mean_absolute_error(y_test, predictions)
mae_1
```
