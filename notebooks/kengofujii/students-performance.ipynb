# Conclusion

The following items were found from this dataset.
* The students who used to have lunch and the standard had a higher average score.
* Students who complete the test preparation course have higher average scores.
* Female have higher reading and writing scores than male.
* Male have higher math score than female.
* When parents' education level is high school, students tend to have low scores.


The results of the top 50 were as follows.
* 70% of the top 50 scorers are female, and 70% of the top 50 scorers completed the test preparation course.
* 90% of the top 50 scorers' parents graduated from college.
* Most of the top scorers ate "standard" lunch. â‡¨ It seems that there is no need to save lunch.

**There is a correlation between the academic ability of students and the educational and economic levels of parents.**
