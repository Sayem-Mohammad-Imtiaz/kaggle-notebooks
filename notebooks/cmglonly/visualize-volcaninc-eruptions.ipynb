## What is the most type of volcanic effect?

```python
df.Type.value_counts()
```

```python
df.Type.value_counts().plot(kind='bar', figsize=(20,8))
```

We could understand that the top three of volcanic types consist of

1. Stratovolcano
2. Shield
3. Stratovolcano(es)

```python
df["Tectonic Setting"].value_counts()
```

```python
df["Tectonic Setting"].value_counts().sort_values().plot(kind='barh')
```

Then, the most common tectonic setting from a volcanic events consist of

1. Subduction Zone / Continental Crust (>25 km)
2. Intraplate / Continental Crust (>25 km)
3. Subduction Zone / Oceanic Crust (< 15 km)

```python
df["Dominant Rock Type"].value_counts()
```

```python
df["Dominant Rock Type"].value_counts().sort_values().plot(kind='barh')
```

Lastly, the most Dominant Rock Type from a volcanic events are
1. Andesite / Basaltic Andesite
2. Basalt / Picro-Basalt
3. Dacite


## Which country has the most active volcanic activity?

```python
df['Country'].value_counts()
```

```python
df['Country'].value_counts().head(10).plot(kind='bar', figsize=(8,6))
```

It seems that US has the most active volcanic activity, followed by Russia, and Indonesia
