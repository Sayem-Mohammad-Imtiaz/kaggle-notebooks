```python
#Renaming lat long colum
WW2data.rename(columns={'Target Latitude':'TLX','Target Longitude':'TLY','Theater of Operations':'Ops'  },inplace=True)
WW2data.head(1)
```

```python
# Cleaning up some NA data
WW2dataX=WW2data[WW2data.TLX.notna()]
WW2datafinal=WW2dataX[WW2dataX.TLY.notna()]
WW2datafinal.isnull().sum()
```

# For Random 10K samples,finding where the bombings took place. Please zoom in to check

```python
WW2finalsample=WW2datafinal.sample(n=10000)
from folium import plugins
worldmap=folium.Map(location=[45.94,24.25],zoom_start=2)

#instantiate a marker cluster
operations = plugins.MarkerCluster().add_to(worldmap)
for lat,lng in zip(WW2finalsample.TLX,WW2finalsample.TLY):
    folium.Marker(
    [lat,lng],fill=True).add_to(operations)
worldmap
```
