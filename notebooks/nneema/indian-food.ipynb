```python
food_df = pd.read_csv("/kaggle/input/indian-food-101/indian_food.csv")
food_df.head()
```

# How many states data do we have?

```python
no_of_state = food_df["state"].unique()
print(no_of_state, len(no_of_state))
```

There is missing rows of information in state with value -1. Let's  drop those rows

```python
print("Before dropping rows",len(food_df))
food_df.drop(food_df[food_df["state"]=="-1"].index, inplace=True)
print("After dropping rows", len(food_df))
```

# Lets check on different flavors of food liked by Indians

```python
import seaborn as sns
```

```python
sns.countplot( x="flavor_profile", data=food_df);
```

Ok.. That's correct people in India likes spicy food.

Lets check which state has more dessert lover

```python
dessert=food_df[food_df["course"]=="dessert"]
len(dessert)
```

```python
top_seller=dessert["state"].value_counts()[0:10]
sns.barplot(y=top_seller.index, x=top_seller.values);
```

# # Let's prove our hypothesis is West Bengal "Rasgulla" lover? :)

```python
if any(dessert[dessert["state"]=="West Bengal"].name.str.contains("Rasgulla")):
    print("Yes!!! They are Rasssgulla lovers")
```
