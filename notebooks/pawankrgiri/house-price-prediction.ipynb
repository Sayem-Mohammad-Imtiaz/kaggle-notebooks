```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
train['BHK_OR_RK']=le.fit_transform(train['BHK_OR_RK'])
train['POSTED_BY']=le.fit_transform(train['POSTED_BY'])
test['POSTED_BY']=le.fit_transform(test['POSTED_BY'])
test['BHK_OR_RK']=le.fit_transform(test['BHK_OR_RK'])
```

```python
train.drop('ADDRESS',axis=1,inplace=True)
train.head()
test.drop('ADDRESS',axis=1,inplace=True)
```

```python
corr = train.corr()
plt.figure(figsize=(30,20))
sns.heatmap(corr,annot=True)
```

```python
corr['TARGET(PRICE_IN_LACS)'].sort_values(ascending=False)
```

```python
x=train.drop(
    'TARGET(PRICE_IN_LACS)',axis=1)
y=train['TARGET(PRICE_IN_LACS)']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.8,random_state=42)
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error
pred = lr.predict(x_test)
print('Mean Squared Error -->',mean_squared_error(y_test,pred))
```

```python
submission=lr.predict(test)
```

```python
pd.DataFrame(submission).to_csv('submission.csv')
```
