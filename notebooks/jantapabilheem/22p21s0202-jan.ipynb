```python
import datetime
import math
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import seaborn as sns
from numpy.polynomial.polynomial import polyfit

#เลือกข้อมูลของโรคระบาด Covid-19 ซึ่งข้อมูลในแต่ละแถวนั้นก็จะรายงานจำนวนผู้ติดเชื้อสะสม จำนวนผู้เสียชีวิต และผู้ป่วยที่หายแล้วแบบอัพเดทในแต่ละวัน

file_url="../input/social-network-ads/Social_Network_Ads.csv"
dataSet = pd.read_csv(file_url)
dataSet.head()
# dataSet.tail()
# print(df)
```

ข้อมูลจากไฟล์จะเห็นว่ามีตัวแปรที่สำคัญต่าง ๆ ชเ่น เพศ อายุ รายได้โดยประมาณ และการตัดสินใจซื้อนั่นเอง

```python
dataSet["Purchased"].value_counts() #เลือกนับค่าในคอลัมล์ชื่อ Purchased โดยค่าที่ได้คือ 1 และ 0
```

```python
dataSet.plot(kind="scatter", x="EstimatedSalary", y="Age")
```

ทำการแบ่งแกน x และ y โดยกำหนดให้ x = ค่าเงินเดือนของผู้ใช้และ y = อายุของผู้ใช้ซึ่งกำหนดให้แสดงค่ากราฟแบบ scatter

```python
sns.jointplot(x="EstimatedSalary", y="Age", data=dataSet, size=5)
```

```python
# One piece of information missing in the plots above is what species each plant is
# We'll use seaborn's FacetGrid to color the scatterplot by species
sns.FacetGrid(dataSet, hue="Purchased", size=5) \
   .map(plt.scatter, "EstimatedSalary", "Age") \
   .add_legend()
```

กราฟ ชนิด FacetGrid

```python
sns.boxplot(x="EstimatedSalary", y="Age", data=dataSet)
```

```python
ax = sns.boxplot(x="EstimatedSalary", y="Age", data=dataSet)
ax = sns.stripplot(x="EstimatedSalary", y="Age", data=dataSet, jitter=True, edgecolor="gray")
```

```python
sns.FacetGrid(dataSet, hue="Purchased", size=6) \
   .map(sns.kdeplot, "Age") \
   .add_legend()
```

```python
sns.pairplot(dataSet.drop("User ID", axis=1), hue="Purchased", size=3)
```

```python
dataSet.drop("EstimatedSalary", axis=1).boxplot(by="Purchased", figsize=(12, 6))
```
