```python
# Directive pour afficher les graphiques dans Jupyter
%matplotlib inline
import numpy as np
# Pandas : librairie de manipulation de données
# NumPy : librairie de calcul scientifique
# MatPlotLib : librairie de visualisation et graphiques
# SeaBorn : librairie de graphiques avancés
import pandas as pd
import seaborn as sns
import xgboost as XGB
from matplotlib import pyplot as plt
from sklearn import ensemble
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
dataset = pd.read_csv("../input/house-price-prediction-challenge/train.csv")
```

```python
dataset = dataset.rename(columns={'TARGET(PRICE_IN_LACS)': 'PRICE_IN_LACS'})
dataset.head()
```

```python
dataset.POSTED_BY.value_counts()
```

# Formatage des champs en chiffres

```python
dataset.POSTED_BY = dataset.POSTED_BY.map({"Owner": 0, "Dealer": 1,"Builder": 2})
dataset.BHK_OR_RK = dataset.BHK_OR_RK.map({"BHK": 0, "RK": 1})
dataset = dataset.drop(["ADDRESS"],axis=1)
dataset.info()
dataset.head()
```

# Vérification de l'intégrité des données

```python
dataset.count()
```

# Formatage des données et découpe des données de test

```python
X = dataset.drop(['PRICE_IN_LACS'], axis=1)
y = dataset.PRICE_IN_LACS
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print(X_train.shape)
print(X_test.shape)
```

# Choix du modèle et entrainement

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
rf.fit(X_train, y_train);
```

# Prédiction des données et calcul de précision

```python
predictions = rf.predict(X_test)
errors = abs(predictions - y_test)
print('Mean Absolute Error:', round(np.mean(errors), 2), 'degrees.')
mape = 100 * (errors / y_test)
accuracy = 100 - np.mean(mape)
print('Accuracy:', round(accuracy, 2), '%.')
```
