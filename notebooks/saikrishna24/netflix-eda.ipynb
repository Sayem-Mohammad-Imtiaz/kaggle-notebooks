## Treating the null values for the most frequent occurrence

```python
def frequent(nfdata,variable,mode):
    nfdata[variable+'_mode']=nfdata[variable].fillna(nfdata[variable].mode()[0])
```

```python
mode=nfdata.country.mode()
```

```python
mode
```

```python
frequent(nfdata,'country',mode)
```

```python
frequent(nfdata,'rating',mode)
```

```python
nfdata.head()
```

## Droping the unwanted coloumns

```python
nfdata.drop('country',inplace=True,axis=1)
```

```python
nfdata.drop('rating',inplace=True,axis=1)
```

```python
nfdata
```

## Treating the null values for Director

### since we Don't Know the director we are filling it will the new catagorical variable

```python
nfdata['director']=np.where(nfdata['cast'].isnull(),'unknown',nfdata['cast'])
```

```python
nfdata
```

## Treating the null values for Date_added

```python
nfdata["date_added"] = pd.to_datetime(nfdata['date_added'])
```

```python
nfdata['date_added'].isnull().sum()
nfdata['month_added']=nfdata['date_added'].dt.month
nfdata['month_name_added']=nfdata['date_added'].dt.month_name()
nfdata['year_added'] = nfdata['date_added'].dt.year
```

```python
nfdata = nfdata.dropna()
```

```python
nfdata.isnull().sum()
```

Till now we have completed treating the null values

# Visualizing the data

## TV show VS Movies

```python
fig = plt.figure()
sns.set(style="whitegrid")
ax = sns.countplot(x="type", data=nfdata, palette="Set1")
```

From the above figure we came to know that total 2000+ Tv show and 4700+ Movies are available in Netflix

```python
plt.figure(figsize=(12,10))
sns.set(style="darkgrid")
ax = sns.countplot(y='release_year', data=nfdata, palette="Set2", order=nfdata['release_year'].value_counts().index[0:15])
```

From the above figure we came to know that the shows are increasing year by year and there is no decreases

## Top 10 Countries useing Netflix

```python
plt.figure(figsize=(12,10))
sns.countplot(y='country_mode',data=nfdata,order=nfdata['country_mode'].value_counts().index[0:15])
```

United States in the top most country using Netflix.

## Top 10 countries using Netflix in percentage


```python
countries = nfdata['country_mode'].value_counts()
list_countries = list(countries.index)
```

```python
list_countries[0:5]
```

```python
plt.figure(figsize=(10, 20))

figure = plt.pie(countries.sort_values(ascending=False).iloc[:10],pctdistance=0.6,labeldistance=1.1,
                 colors='rgb',labels=list_countries[:10],autopct='%1.2f%%')
```

From the pie chart we came to know that the United States uses more Netflix other then the other Countries
