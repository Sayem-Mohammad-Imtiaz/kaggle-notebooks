```python
train_file_path = '../input/heart-attack-analysis-prediction-dataset/heart.csv'

heart_train_data = pd.read_csv(train_file_path)

heart_train_data.describe(include='all')
```

```python
heart_train_data.columns
# Displays colums from the data set
```

# Definitions of colums

- age - Age
- sex - Sex (its either a one or zero but the actual gender for the 1 or 0 has not been provided)
- cp - chest pain type ~ 0 = typical angina(chest discomfort or shortness of breath.), 1 = atypical angina, 2 = non-anginal Pain(very minor) 3 = aymptomatic(no symptoms)
- trtbps - resting blood pressure 
- chol - cholestoral 
- fbs - (fasting blood sugar > 120 mg) ~ 1 = True, 0 = False
- restecg - resting eletrical activity in the heart results ~ 0 = Normal, 1 = ST-T wave normality, 2 = Left ventricular hypertrophy
- thalachh - maximum heart rate achieved
- oldpeak - previous peak
- slp - slope ( no given info about this one )
- caa -number of major vessels



```python
selected_columns = ['age','sex','trtbps','chol','fbs','thalachh','caa','restecg','cp','output','oldpeak',]

prepared_data = heart_train_data[selected_columns]

prepared_data = prepared_data.dropna(axis=0)

prepared_data.describe(include='all')
```

Removed slope since no actual information was providied for what it was, and going through the data didn't have any significant values or change so I removed it, additionally it didnt fit in with the code being used.

```python
# Seperates out the prediction target
y = prepared_data.output

# drops the original dataframe and use the rest as feature data
X = prepared_data.drop('output', axis=1)

X.head()
```

```python
# Amplified the max depth to 3 for easy display
attack_predictor = DecisionTreeClassifier(max_depth=3)

# Trains the model
attack_predictor.fit(X, y)
# Plots the tree
plt.figure(figsize = (20,10))
plot_tree(attack_predictor,
          feature_names=heart_train_data.columns,
          class_names=['high chance', 'low chance'],
          filled=True)
plt.show()
```

##### Tree
If i used a Decision tree regressor i would of received a numeric value rather than a classification, since im trying to figure out what class it is (high or low chance) I need a catagorised 1 and 0 rather than a numeric value without much purpose for this model.

# Predicitons

```python
# Get the first five predictions as a list
pred = attack_predictor.predict(X)

print("The predictions are:")
# Merges both target values and predicition values \
X['Prediction'] = pred

X.head()
```

Judging by the displayed data and its predictions it seems to be working succesfully, now the data will be tested to prove its accuracy score through a logisitic regressor. With this I can verify the models accuracy.

## Logistic regression

In this it will use the colums selected and split the data into training and testing data and then use a accuracy score for the logistic regressor. Using this will be able to get a number value for the accuracy of are new trained data.

```python
# splitting data into training data and testing data
from sklearn.model_selection import train_test_split
X = prepared_data.drop(columns=['output'])
y = prepared_data['output']
# stratify arranges and classifiys.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression #import Logistic regressor and then fit the data

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Uses a model score test accuracy and create a number
predicted = model.predict(X_test)
model.score(X_test, y_test)
```

```python
# reword the score into a sentence 
log_ac=accuracy_score(y_test, predicted)*100
print ("The accuracy of Logistic Regression is : ", log_ac, "%")
```
