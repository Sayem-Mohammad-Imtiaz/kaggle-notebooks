```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.utils import shuffle

sns.set()
%matplotlib inline
```

```python
df = shuffle(pd.read_csv("/kaggle/input/social-power-nba/nba_2017_nba_players_with_salary.csv"))
df = df.drop(columns=['PLAYER', 'POSITION', 'TEAM', 'Rk', 'Unnamed: 0']).dropna()
```

```python
#80% of the data is training and 20% is a test data set
train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
```

```python
#Copy of training dataset for data exploration 
df_copy = train_set.copy()

# Basic analysis of dataset
df_copy.describe()
```

```python
df_corr = df_copy.corr()
df_corr
```

```python
train_set_full = train_set.copy()

# Drop salary info from training set. Trying to evaluate what factors cause a higher salary
train_set = train_set.drop(["SALARY_MILLIONS"], axis=1, errors='ignore')
```

```python
# For comparissons
train_labels = df_copy["SALARY_MILLIONS"]
```

```python
lin_reg = LinearRegression()

lin_reg.fit(train_set, train_labels)
```

```python
coefficients = lin_reg.coef_
test_set_cols = train_set.columns
list(zip(test_set_cols, coefficients)) 
```

```python
test_set_no_salary = test_set.drop(columns=['SALARY_MILLIONS'])
```

```python
salary_pred = lin_reg.predict(test_set_no_salary)

#test salary to actual salary map
list(zip(salary_pred, test_set['SALARY_MILLIONS']))
```

```python
lin_reg.score(test_set_no_salary, test_set["SALARY_MILLIONS"])
```
