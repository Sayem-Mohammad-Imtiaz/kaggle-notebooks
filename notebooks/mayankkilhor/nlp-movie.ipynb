```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("../input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv")
```

```python
dataset.dtypes
```

```python
import re
import nltk
```

```python
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()

data = []
```

```python
dataset.shape
```

```python
for i in range(0,50000):
    review = dataset["review"][i]
    review = re.sub('[^a-zA-Z]',' ',review)
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    data.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 2000)
x = cv.fit_transform(data).toarray()
```

```python
x
```

```python
y = dataset.iloc[:,1:2].values 
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([("oh",OneHotEncoder(),[0])], remainder="passthrough")
y = ct.fit_transform(y)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2, random_state = 0)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model =  Sequential()
```

```python
x_train.shape
```

```python
model.add(Dense(units = 2000,kernel_initializer = "random_uniform",activation= "relu"))
```

```python
model.add(Dense(units = 4000,kernel_initializer = "random_uniform",activation= "relu"))
```

```python
model.add(Dense(units = 4000,kernel_initializer = "random_uniform",activation= "relu"))
```

```python
model.add(Dense(units = 2,kernel_initializer = "random_uniform",activation= "softmax"))
```

```python
model.compile(optimizer = "adam",loss = "categorical_crossentropy",metrics = ["accuracy"])
```

```python
model.fit(x_train,y_train,epochs=5)
```

```python
model.save("movieReview.h5")
```
