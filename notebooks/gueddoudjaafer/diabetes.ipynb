```python
Features = data.drop('Outcome', axis=1)
target = data['Outcome']

from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(Features, target, test_size=0.3, random_state=1)
```

```python
#Check Zero in colums

print("Total number of rows: {}".format(len(data)))
print("Total of rows missing in Pregnancies column: {}".format(len(data.loc[data['Pregnancies']==0])))
print("Total of rows missing in Glucose column: {}".format(len(data.loc[data['Glucose']==0])))
print("Total of rows missing in Blood Pressure column: {}".format(len(data.loc[data['BloodPressure']==0])))
print("Total of rows missing in Skin Thickness column: {}".format(len(data.loc[data['SkinThickness']==0])))
print("Total of rows missing in Insulin column: {}".format(len(data.loc[data['Insulin']==0])))
print("Total of rows missing in BMI column: {}".format(len(data.loc[data['BMI']==0])))
print("Total of rows missing in Diabetes Pedigree Function column: {}".format(len(data.loc[data['DiabetesPedigreeFunction']==0])))
print("Total of rows missing in Age column: {}".format(len(data.loc[data['Age']==0])))
```

```python
from sklearn.impute import SimpleImputer

fill_values = SimpleImputer(missing_values=0, strategy='mean')
xtrain = fill_values.fit_transform(xtrain)
xtest = fill_values.fit_transform(xtest)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix

model1 = RandomForestClassifier()
model1.fit(xtrain, ytrain)
model1_pred = model1.predict(xtest)
print('Random Forest Accuracy score: {}'.format(metrics.accuracy_score(ytest, model1_pred)))
print(confusion_matrix(ytest, model1_pred))
print(classification_report(ytest, model1_pred))
```

```python
model2= DecisionTreeClassifier()
model2.fit(xtrain, ytrain)
model2_pred = model2.predict(xtest)
print('Decision Tree Accuracy score: {}'.format(metrics.accuracy_score(ytest, model2_pred)))
print(confusion_matrix(ytest, model2_pred))
print(classification_report(ytest, model2_pred))
```

```python
model3 = XGBClassifier(gamma=0)
model3.fit(xtrain, ytrain)
model3_pred = model3.predict(xtest)
print('XGBoost Accuracy score: {}'.format(metrics.accuracy_score(ytest, model3_pred)))
print(confusion_matrix(ytest, model3_pred))
print(classification_report(ytest, model3_pred))
```

```python
xgb_features = model3.feature_importances_
(pd.Series(xgb_features, index=Features.columns).plot(kind='barh'))
```
