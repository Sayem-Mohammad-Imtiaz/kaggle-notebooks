```python
#Convert str to datetime object
nt['date_added'] = pd.to_datetime(nt['date_added'])
```

```python
nt['date_added'].iloc[3]
```

```python
time = nt['date_added'].iloc[200]
```

```python
nt['added_month'] = nt['date_added'].apply(lambda time: time.month)
nt['added_day'] = nt['date_added'].apply(lambda time: time.dayofweek)
nt['added_year'] = nt['date_added'].apply(lambda time: time.year)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
# mmap = {1.0: 'Jan', 2.0: 'Feb', 3.0: 'Mar', 4.0: 'Apr', 5.0: 'May',
#             6.0: 'Jun', 7.0: 'Jul', 8.0: 'Aug', 9.0: 'Sep', 10.0: 'Oct', 11.0: 'Nov', 12.0: 'Dec'}
```

```python
nt['added_day'] = nt['added_day'].map(dmap)
#nt_movies['added_month'] = nt_movies['added_month'].map(mmap)
nt.head()
```

```python
plt.figure(figsize=(20,6))
sns.countplot(x='added_year', data=nt, palette='viridis',)
```

```python
#Movies
monthYear = nt.groupby(by=['added_month','added_year']).count()['show_id'].unstack().fillna(0)
monthYear
```

```python
plt.figure(figsize=(20,12))
sns.heatmap(monthYear,annot=True, cmap="coolwarm", fmt='g')
```
