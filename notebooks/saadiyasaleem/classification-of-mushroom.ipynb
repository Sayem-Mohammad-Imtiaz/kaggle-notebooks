# CLASSIFICATION OF MUSHROOMS.

*THIS NOTEBOOK GIVES FEW MACHINE LEARNING MODELS FOR CLASSIFYING THE DIFFERENT TYPES OF MUSHROOM.*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

READING THE DATA

```python
df = pd.read_csv("../input/mushroom-classification/mushrooms.csv")
```

```python
df.head()
```

```python
df.info()
```

DATA PRE PROCESSING 

```python
from sklearn.preprocessing import LabelEncoder

for col in df.columns: # Consider each column (except 'y' column, we deal with that separately)
    
    # Generate labels for each value in column
    le = LabelEncoder();
    le.fit(df[col].unique());
    
    # Print labels
    print('\n' + col + ':');
    for val in df[col].unique():
        print(val, int(le.transform([val])));
        
    # Transform data    
    df[col] = le.transform(df[col]);
    
print('\n Sample of transformed data-frame:');
```

```python
df.head()
```

```python
X=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=0)
```

Machine learning

neural networks

```python
from sklearn.neural_network import MLPClassifier

nn = MLPClassifier().fit(X_train, y_train);
```

SVMs

```python
from sklearn.svm import SVC

svm = SVC(gamma = 'auto').fit(X_train, y_train);
```

Decision trees

```python
from sklearn import tree

dt = tree.DecisionTreeClassifier().fit(X_train, y_train);
```

Accuracy of models

```python
print("Accuracy for nn");
print("Accuracy on train-set:", "%.2f" %(nn.score(X_train, y_train)*100), '%');
print("Accuracy on test-set:", "%.2f" %(nn.score(X_test, y_test)*100), '%');
print("Accuracy for svm");
print("Accuracy on train-set:", "%.2f" %(svm.score(X_train, y_train)*100), '%');
print("Accuracy on test-set:", "%.2f" %(svm.score(X_test, y_test)*100), '%')
print("Accuracy for dtc");
print("Accuracy on train-set:", "%.2f" %(dt.score(X_train, y_train)*100), '%');
print("Accuracy on test-set:", "%.2f" %(dt.score(X_test, y_test)*100), '%');
```

Random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
rand_clf = RandomForestClassifier(random_state=0)
rand_clf.fit(X_train,y_train)

rand_clf.score(X_test,y_test)
```
