# 1)Linear Regression
mathematical formula :

y = b0 + b1*x

y = Dependent variable 

x = Independent variable

b0 = Costant variable(bias)

b1 = Weight of x(slope)

A feature (x) is required for linear regression so I used fixed acidity column for this regression. And target is pH column.

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv")
x = np.array(df.loc[:,"fixed acidity"]).reshape(-1,1)
y = np.array(df.loc[:,"pH"]).reshape(-1,1)
plt.figure(figsize=[25,10])
plt.scatter(x,y)
plt.xlabel("Fixed Acidity",fontsize=28)
plt.ylabel("pH",fontsize=28)
plt.show()
```

```python
#%% prediction
from sklearn.linear_model import LinearRegression

Lreg = LinearRegression()

Lreg.fit(x,y)
y_head=Lreg.predict(x).reshape(-1,1)

b0=Lreg.predict([[0]])
b1=Lreg.coef_
print("b0 : " , b0)
print("b1 : ", b1)
# predict ph when fixed acidity equal 16
print("Predict  : " , (b0+(b1*16)))
```

```python
#%% R^2 score
from sklearn.metrics import r2_score
print("R^2 : ",r2_score(y,y_head))

```

```python
#%% # Plot regression line and scatter
plt.figure(figsize=[25,10])
plt.scatter(x,y)
plt.xlabel("Fixed Acidity",fontsize=28)
plt.ylabel("pH",fontsize=28)
plt.plot(x,y_head,color="red",linewidth=4)
plt.show()
```

# 2)Multiple Linear Regression

mathematical formula :

y=b0+b1x1+b2x2+...+bn*xn

In multiple linear regression we have more features. I used all the features in the csv file 
    

```python

from sklearn.linear_model import LinearRegression

x = df.iloc[:,[0,1,2,3,4,5,6,7,9]].values
y = df.iloc[:, 0].values.reshape(-1,1)
multiple_Lreg=LinearRegression()

multiple_Lreg.fit(x,y)
b0 = multiple_Lreg.intercept_
bs = multiple_Lreg.coef_
y_head=multiple_Lreg.predict(x)
print("b0 : " ,b0)
dict_of_b={}
z=1
for i in bs:
    for j in i:
        if z==8:
            dict_of_b["b"+str(z+1)]=j
            continue
        dict_of_b["b"+str(z)]=j
        
        z+=1

print("Weight of all x : ",dict_of_b)

```

```python
#%% r2 score
from sklearn.metrics import r2_score
print("R^2 : ",r2_score(y,y_head))

```

# 3)Polynomial Regression

mathematical formula :

y=b0+b1x+b2x^2+...+bn*x^n


```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_regression = PolynomialFeatures(degree = 3)
x_polynomial = polynomial_regression.fit_transform(x)
```

```python
from sklearn.linear_model import LinearRegression
plt.figure(figsize=[25,10])

poly_reg=LinearRegression()
poly_reg.fit(x_polynomial,y)
y_head=poly_reg.predict(x_polynomial)
plt.plot(x,y_head,color="purple")
plt.show()
```

```python
#%% R^2 score
from sklearn.metrics import r2_score
print("R^2 : ",r2_score(y,y_head))
```

# 4)Decision Tree Regression


```python
from sklearn.tree import DecisionTreeRegressor

x = df.loc[:,"fixed acidity"].values.reshape(-1,1)
y = df.loc[:,"pH"].values.reshape(-1,1)

dTree = DecisionTreeRegressor()
dTree.fit(x,y)
x_ =  np.arange(min(x),max(x),0.001).reshape(-1,1)
y_head = dTree.predict(x_)
```

```python
plt.figure(figsize=[25,10])
plt.scatter(x,y,color="blue")
plt.plot(x_,y_head,color="red" ,linewidth=6)
plt.xlabel("Fixed Acidity" ,fontsize=25)
plt.ylabel("pH" ,fontsize=25)
plt.show()
```

# 5)Random Forest Regression



```python


from sklearn.ensemble import RandomForestRegressor

rf=RandomForestRegressor(n_estimators=200,random_state=(42))

rf.fit(x,y)
x_=np.arange(x.min(),x.max(),0.001).reshape(-1,1)

y_head=rf.predict(x_).reshape(-1,1)
```

```python
plt.figure(figsize=[25,10])
plt.scatter(x,y,color="brown")
plt.plot(x_,y_head,color="green",linewidth=3)
plt.xlabel("Fixed Acidity")
plt.ylabel("pH")
plt.show()
```
