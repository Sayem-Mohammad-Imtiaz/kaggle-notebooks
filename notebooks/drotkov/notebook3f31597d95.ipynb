```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
df = pd.read_csv('../input/housesalesprediction/kc_house_data.csv')
```

3. Изобразите распределение цен на дома (столбец price) в виде гистограммы. Подберите количество интервалов (bins) экспериментально. Добавьте подпись к оси Ox и название графика.

```python
plt.title('Распределение цен')
plt.xlabel('Price')
plt.hist(x=df['price'], bins=100)
plt.show()
```

4. Постройте график для визуализации соотношения общей жилой площади (sqft_living) и цены (price). Задайте параметры графика:
• подписи к осям, название графика;
• тип маркера: circle; размер: 20; цвет: зелёный.

```python
plt.plot(df['price'], df['sqft_living'], 'go', markersize=2)
plt.title("Соотношение цены и жилой площади")
plt.xlabel("Цена")
plt.ylabel("Общая жилая площадь")
plt.show()
```

5. Дополните параметры предыдущего графика, чтобы показать наличие/отсутствие вида на море (столбец waterfront). Сделайте разбиение на группы по признаку waterfront путём разделения на два графика. Какая из двух последних визуализаций нагляднее доносит информацию?

```python
filt1 = df['waterfront'] == 1
filt0 = df['waterfront'] == 0
plt.plot(df.loc[filt0]['price'], df.loc[filt0]['sqft_living'], 'go',
         df.loc[filt1]['price'], df.loc[filt1]['sqft_living'], 'bx',
         markersize=2)
plt.title("Соотношение цены и жилой площади")
plt.xlabel("Цена")
plt.ylabel("Общая жилая площадь")
plt.show()
```

```python
plt.subplot(2, 1, 1)
plt.plot(df.loc[filt0]['price'], df.loc[filt0]['sqft_living'], 'bo', markersize=2)
plt.title("Соотношение цены и жилой площади с видом на море")
plt.xlabel("Цена")
plt.ylabel("Общая жилая площадь")

plt.subplot(2, 1, 2)
plt.plot(df.loc[filt1]['price'], df.loc[filt1]['sqft_living'], 'go', markersize=2)
plt.title("Соотношение цены и жилой площади без вида на море")
plt.xlabel("Цена")
plt.ylabel("Общая жилая площадь")
plt.show()
```

Первая визуализация видится более наглядной.
