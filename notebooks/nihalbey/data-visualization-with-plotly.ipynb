```python
#We will analyze just times data
data = pd.read_csv("../input/timesData.csv")
```

```python
#Firstly we know what the data and what is there in data
data.info()
```

```python
data.head()
```

**Line Chart**

```python
#Citation and Teaching vs World Rank of Top 100 Universities
#we can draw good plot for this question with line plot(chart)
top_100 = data.iloc[:100,:]
trace1 = go.Scatter(
x=top_100.world_rank,
y=top_100.citations,
mode = "lines",
marker={"color":"rgba(48, 64, 101, 0.8)"},
text= top_100.university_name)
trace2= go.Scatter(
x=top_100.world_rank,
y=top_100.teaching,
mode = "lines+markers",
marker=dict(color="rgba(0,123,221,.8)"),
text=top_100.university_name)
data1 = [trace1,trace2]
layout = dict(title = "Citition and Teaching vs World Rank", xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False))
fig = dict(data = data1, layout = layout)
iplot(fig)
```

**Scatter Chart**

```python
#Citation vs world rank of top 100 universities with 2014, 2015 and 2016 years
x2014 = data[data.year == 2014].iloc[:100,:]
x2015 = data[data.year == 2015].iloc[:100,:]
x2016 = data[data.year == 2016].iloc[:100,:]
trace1 =go.Scatter(
                    x = x2014.world_rank,
                    y = x2014.citations,
                    mode = "markers",
                    name = "2014",
                    marker = dict(color = 'rgba(255, 128, 255, 0.8)'),
                    text= x2014.university_name)
# creating trace2
trace2 =go.Scatter(
                    x = x2015.world_rank,
                    y = x2015.citations,
                    mode = "markers",
                    name = "2015",
                    marker = dict(color = 'rgba(255, 128, 2, 0.8)'),
                    text= x2015.university_name)
# creating trace3
trace3 =go.Scatter(
                    x = x2016.world_rank,
                    y = x2016.citations,
                    mode = "markers",
                    name = "2016",
                    marker = dict(color = 'rgba(0, 255, 200, 0.8)'),
                    text= x2016.university_name)
data1 = [trace1, trace2, trace3]
layout = dict(title = 'Citation vs world rank of top 100 universities with 2014, 2015 and 2016 years',
              xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'Citation',ticklen= 5,zeroline= False)
             )
fig = dict(data = data1, layout = layout)
iplot(fig)
```

**Bar Chart**

```python
#citations and teaching of top 3 universities in 2014 
#We draw bar plot for showing citations and teaching top 3 universities
x2014 = data[data.year == 2014].iloc[:3,:]
trace1 = go.Bar(
x = x2014.university_name,
y = x2014.citations,
name="citiations",
marker = {"color":"rgba(111,23,155,0.5)"},
text=x2014.university_name
)
trace2 = go.Bar(
x = x2014.university_name,
y = x2014.teaching,
name="teaching",
marker = {"color":"rgba(47,69,187)"},
text = x2014.university_name
)
data3 = [trace1,trace2]
layout = go.Layout(barmode = "group")
fig = go.Figure(data = data3, layout = layout)
iplot(fig)
```

```python
#We draw bar plot differt type.This is relative bar plot 
layout = go.Layout(barmode = "relative")
fig = go.Figure(data = data3,layout = layout)
iplot(fig)
```

```python
#this my questions land belong the most number of university on rank in 2015 and 2016
#Already we have 2016 years data
index = x2016["country"].value_counts().head(3).index
value = x2016["country"].value_counts().head(3).values
trace1 = go.Bar(
x = index,
y = value,
marker = {"color":"rgba(131,26,93,0.4)"}    
)
data4 = [trace1]

iplot(data4)
```

```python
x2016 = data[data.year == 2016].iloc[:20,:]
num_students_size  = [float(each.replace(',', '.')) for each in x2016.num_students]
international_color = [float(each) for each in x2016.international]
trace1 = go.Scatter(
x = x2016.world_rank,
y = x2016.teaching,
mode = "markers",
marker=dict(
color = international_color,
size = num_students_size,
showscale = True
),
text = x2016.university_name
)
data5 = [trace1]
iplot(data5)
```

**3D Scatter plot**

we do 3d plot but we add extra dimension with color and size

```python
# we want to my plots x axis world rank, y axis teaching,z axis citiation,color is research,size is total score in 2016
color_3d = x2016.research 
x2016.total_score = x2016.total_score.astype(float)
size_3d = x2016.total_score
trace = go.Scatter3d(
x = x2016.world_rank,
y = x2016.teaching,
z = x2016.citations,
mode = "markers",
marker = dict ( color = color_3d,size = 10),
text = x2016.university_name 
)
data6 = [trace]
layout = go.Layout(margin = dict(l=0,r=0,b=0,t=0))
fig = go.Figure(data = data6,layout = layout)
iplot(fig)
```

if you add 5th dimension you can do size = someones numeric value ex: size = data.total_score(total_score must be numeric)

**Inset Plot**

if we wanto se two plot example scatter + line or line + histogram.We use this plot.

```python
#we draw plot has scatter plot + bar plot
x2016 = data[data.year == 2016]
trace1 = go.Bar(
x = x2016.world_rank,
y = x2016.teaching,
marker = dict(color = "rgba(25,63,86,0.5)"),
xaxis = "x2",
yaxis ="y2",
name = "t"
)
trace2 = go.Scatter(
x = x2016.world_rank,
y = x2016.citations,
mode = "markers",
marker = dict (color = "rgba(52,65,144,0.55)"),
text = x2016.world_rank,
name = "c"
)
data7 = [trace1,trace2]
layout = go.Layout(
xaxis2=dict(
        domain=[0.6, 0.95],
        anchor='y2',        
    ),
    yaxis2=dict(
        domain=[0.6, 0.95],
        anchor='x2',
    ),
title = "teaching and citations vs university rank"
)
fig = go.Figure(data = data7,layout = layout)
iplot(fig)
```

```python
data.total_score.replace("-",10,inplace = True)
data.total_score = data.total_score.astype(float)
```

```python
#one more inset plot.This plot is line + box plot (it is be box plot)
top_100 = data.iloc[:100,:]
trace1 = go.Scatter(
x = top_100.world_rank,
y = top_100.citations,
mode = "lines",
marker = dict(color = "rgba(200,140,61,0.4)"),
text = top_100.university_name
)
trace2 = go.Box(
y = top_100.total_score,
name = "total score",
marker = dict(color = "rgba(42,155,204,0.6)"),
xaxis = "x2",
yaxis = "y2" ,
)
data9 = [trace1,trace2]
layout = go.Layout(
 xaxis2=dict(
        domain=[0.6, 0.95],
        anchor='y2',        
    ),
    yaxis2=dict(
        domain=[0.6, 0.95],
        anchor='x2',
    ),
title = "researching and teaching"
)
fig = go.Figure(data = data9,layout = layout)
iplot(fig)
```

**Multiple plot**

We draw different content or consept plot in one figure.we can see on multiple plot different feature's plot.

```python
#We comparing citations,teaching,research and total score in same frame 
trace0 = go.Scatter(
x = top_100.world_rank,
y = top_100.teaching,
mode = "lines",
name = "teaching",
marker = dict(color = 'rgba(12, 12, 140,.4)'),
text = top_100.university_name
)
trace1 = go.Scatter(
x = top_100.world_rank,
y = top_100.citations,    
mode ="lines + markers",
name = "citation",
marker = dict(color = "rgba(155,98,160,.6)"),
xaxis = "x2",
yaxis = "y2",
text = top_100.university_name
)
trace2 = go.Scatter(
x = top_100.world_rank,
y= top_100.total_score,
mode = "lines",
name = "total",
marker = {"color":"rgba(36,120,153,.4)"},
xaxis = "x3",
yaxis = "y3",
text = top_100.university_name
)
trace3 = go.Scatter(
x = top_100.world_rank,
y = top_100.research,
mode = "lines + markers",
name = "research",
marker = {"color":"rgba(65,46,178,0.4)"},
xaxis = "x4",
yaxis = "y4",
text = top_100.university_name
)
data9 = [trace0,trace1,trace2,trace3]
layout = go.Layout(
    xaxis=dict(
        domain=[0, 0.45]
    ),
    yaxis=dict(
        domain=[0, 0.45]
    ),
    xaxis2=dict(
        domain=[0.55, 1]
    ),
    xaxis3=dict(
        domain=[0, 0.45],
        anchor='y3'
    ),
    xaxis4=dict(
        domain=[0.55, 1],
        anchor='y4'
    ),
    yaxis2=dict(
        domain=[0, 0.45],
        anchor='x2'
    ),
    yaxis3=dict(
        domain=[0.55, 1]
    ),
    yaxis4=dict(
        domain=[0.55, 1],
        anchor='x4'
    ),
    title = "Multiple"
)
fig = go.Figure(data = data9,layout = layout)
iplot(fig)
```

**Scatter Matrix Plots**

if we want to see relation inter two columns and box plot with table. We use scatter matrix plots

```python
#Firstly we importing library.
import plotly.figure_factory as ff
df2015 = x2015.loc[:,["teaching","citations","research"]]
df2015["index"] = np.arange(1,len(df2015)+1)
fig = ff.create_scatterplotmatrix(df2015, diag = "box",index = "
                                  ",colormap = "Portland", colormap_type='cat',
                                  height=700, width=700)
iplot(fig)
```
