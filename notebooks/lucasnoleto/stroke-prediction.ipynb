```python
'''Attribute Information
1) id: unique identifier
2) gender: "Male", "Female" or "Other"
3) age: age of the patient
4) hypertension: 0 if the patient doesn't have hypertension, 1 if the patient has hypertension
5) heart_disease: 0 if the patient doesn't have any heart diseases, 1 if the patient has a heart disease
6) ever_married: "No" or "Yes"
7) work_type: "children", "Govt_jov", "Never_worked", "Private" or "Self-employed"
8) Residence_type: "Rural" or "Urban"
9) avg_glucose_level: average glucose level in blood
10) bmi: body mass index
11) smoking_status: "formerly smoked", "never smoked", "smokes" or "Unknown"*
12) stroke: 1 if the patient had a stroke or 0 if not
*Note: "Unknown" in smoking_status means that the information is unavailable for this patient'''
```

```python
training_predictors = pd.get_dummies(X_train)
#train_val = pd.get_dummies(y_train)
test_predictors = pd.get_dummies(X_val)
#val_predictors = pd.get_dummies(y_val)
```

```python
tab = ["DecisionTreeRegressor","KNeighborsClassifier","Naive Bayes","Logistic Regression","Decision Tree", "KNN", "Random Forest Classifier","SVC"]
val = []
```

```python
y_val
```

```python
test_predictors['stroke'] = y_val
```

```python
test_predictors
```

```python
# Decision Tree Regression
stroke_model = DecisionTreeRegressor(random_state =1 )


# Fit iowa_model with the training data.
stroke_model.fit(training_predictors, y_train)
from sklearn.metrics import mean_absolute_error

predicted_patients= stroke_model.predict(test_predictors)
mean_absolute_error(y_val, predicted_patients)
val.append(mean_absolute_error(y_val, predicted_patients))
```

```python
#KNeighbors Classifier
knn = KNeighborsClassifier()

knn.fit(training_predictors, y_train)


predicted_patients= knn.predict(test_predictors)
mean_absolute_error(y_val, predicted_patients)
val.append(mean_absolute_error(y_val, predicted_patients))
```

```python
#Naive Bayes 
gnb = GaussianNB()

gnb.fit(training_predictors, y_train)

predicted_patients= gnb.predict(test_predictors)
mean_absolute_error(y_val, predicted_patients)
val.append(mean_absolute_error(y_val, predicted_patients))
```

```python
#Logistic Regression
lr = LogisticRegression(max_iter = 2000)

lr.fit(training_predictors, y_train)

predicted_patients= lr.predict(test_predictors)
mean_absolute_error(y_val, predicted_patients)
val.append(mean_absolute_error(y_val, predicted_patients))
```

```python
#Decision Tree 
dt = tree.DecisionTreeClassifier(random_state = 1)

dt.fit(training_predictors, y_train)

predicted_patients= dt.predict(test_predictors)
mean_absolute_error(y_val, predicted_patients)
val.append(mean_absolute_error(y_val, predicted_patients))
```

```python
#KNN

knn = KNeighborsClassifier()

knn.fit(training_predictors, y_train)

predicted_patients= knn.predict(test_predictors)
mean_absolute_error(y_val, predicted_patients)
val.append(mean_absolute_error(y_val, predicted_patients))

```

```python
#Random Forest Classifier

rf = RandomForestClassifier(random_state = 1)

rf.fit(training_predictors, y_train)
predicted_patients= rf.predict(test_predictors)
mean_absolute_error(y_val, predicted_patients)
val.append(mean_absolute_error(y_val, predicted_patients))
```

```python
#SVC

svc = SVC(probability = True)

svc.fit(training_predictors, y_train)

predicted_patients= svc.predict(test_predictors)
mean_absolute_error(y_val, predicted_patients)
val.append(mean_absolute_error(y_val, predicted_patients))
```
