```python
from sklearn.model_selection import train_test_split
y = data.fetal_health
X = data.drop(['fetal_health'], axis = 1)
X_train, X_val, y_train, y_val = train_test_split(X, y,random_state = 0)
```

```python
X
```

```python
tab = ["KNeighborsClassifier","Naive Bayes","Decision Tree"]
val = []
```

```python
from sklearn.metrics import mean_absolute_error
knn = KNeighborsClassifier()

knn.fit(X_train, y_train)


predicted_patients= knn.predict(X_val)
mean_absolute_error(y_val, predicted_patients)
val.append(mean_absolute_error(y_val, predicted_patients))
```

```python
gnb = GaussianNB()

gnb.fit(X_train, y_train)

predicted_patients= gnb.predict(X_val)
mean_absolute_error(y_val, predicted_patients)
val.append(mean_absolute_error(y_val, predicted_patients))
```

```python
#Decision Tree 
dt = tree.DecisionTreeClassifier(random_state = 1)

dt.fit(X_train, y_train)

predicted_patients= dt.predict(X_val)
mean_absolute_error(y_val, predicted_patients)
val.append(mean_absolute_error(y_val, predicted_patients))
```

```python
validation=list(zip(tab,val))
```
