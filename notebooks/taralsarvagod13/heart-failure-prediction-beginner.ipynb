# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split

X = df.drop('DEATH_EVENT', axis=1).values
y = df.iloc[:,-1]
X
```

# MinMax Scalar

We need to scale the features as some features are in different scale. I have used MinMax Scalar.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_X = pd.DataFrame(scaler.fit_transform(X))
df_X
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_X, y, test_size=0.2, random_state=42)
```

```python
RFC = RandomForestClassifier()
RFC.fit(X_train, y_train)
```

```python
y_pred = RFC.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
```

So, here we can see without doing any feature engineering we can see accuracy of the model is good. It will further improve further once we perform feature engineering, hyperparameter optimization.
