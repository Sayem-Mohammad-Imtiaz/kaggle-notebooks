```python
def f(*args):
    for i in args:
        print(i)
f(2)
print("=======")
f(1,55,22,11)
print("_______________________")
def funk(**kwargs):
    for key,value in kwargs.items():
        print(key,"model is",value)
funk(BWM="M4",Mercedes="Brabus",Lambogini="Gallardo")
```

> Lambda Function****

We are using Lambda Funciton for write function which is easier than normal function writing

```python
square_root=lambda x:x**1/2
print(square_root(4))

print("=========")
sum=lambda x:x+x
print(sum(123))
```

* > But for more complicated functions we use anonymous functions

```python
citylist=["İstanbul","London","New Delhi"]
y=map(lambda x:x+" city",citylist)
print(list(y))

print("or diffent example")

numberlist=[99,11,23,float(0.54)]
a=map(lambda o:o+12,numberlist)
print(list(a))
```

Zip and Unzip on Lists

```python
list1=[99,88,77,66]
list2=[4,3,2,1]
zipp=zip(list1,list2)
print(zipp)
zipplist=list(zipp)
print(zipplist)
```

```python
unzip=zip(*zipplist)
newlist1,newlist2=list(unzip)
print(newlist1)
print(newlist2)
print(type(newlist1))
```
