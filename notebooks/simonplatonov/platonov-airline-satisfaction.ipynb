```python
import pandas as pd

df_train = pd.read_csv('../input/airline-passenger-satisfaction/train.csv')
df_test = pd.read_csv('../input/airline-passenger-satisfaction/test.csv')
```

```python
df_train['satisfaction'] = df_train['satisfaction'].map({'neutral or dissatisfied' : 0, 'satisfied' : 1})

df_test['satisfaction'] = df_test['satisfaction'].map({'neutral or dissatisfied' : 0, 'satisfied' : 1})
```

```python
def one_hot_encode(df, feature):
    one_hot = pd.get_dummies(df[feature], prefix=feature)
    df = df.drop(feature, axis=1)
    return df.join(one_hot)
```

```python
def transform(x):
    df = x.copy()
    df = one_hot_encode(df, 'Gender')
    df = one_hot_encode(df, 'Customer Type')
    df = one_hot_encode(df, 'Type of Travel')
    df = one_hot_encode(df, 'Class')
    df['Arrival Delay in Minutes'].fillna(df['Arrival Delay in Minutes'].median(), inplace = True)
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    df.pop('id')
    df.pop('satisfaction')
    return df
```

```python
x = transform(df_train)
y = df_train.satisfaction
x_test = transform(df_test)
y_test = df_test.satisfaction
```

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size = 0.2, random_state=42)
```

```python
model = XGBClassifier(max_depth = 5, learning_rate = 0.1, n_estimators = 325)

#model.fit(x, y)

#y_pred = model.predict(x_test)

#print("acc_score = ",accuracy_score(y_pred,y_test))

model.fit(x_train, y_train)

y_pred = model.predict(x_val)

print("acc_score = ",accuracy_score(y_pred,y_val))

y_pred = model.predict(x_test)

print("acc_score = ",accuracy_score(y_pred,y_test))
```
