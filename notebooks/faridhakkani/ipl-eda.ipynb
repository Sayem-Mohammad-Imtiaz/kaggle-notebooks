```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ipl = pd.read_csv('../input/ipl-data-set/matches.csv')
```

```python
ipl.head(10)
```

```python
ipl.shape
```

```python
ipl.isnull().values.any()
```

```python
ipl.isnull().sum()
```

```python
ipl.fillna(method='ffill')
```

```python
plt.figure(figsize=(20,6))
sns.countplot(data=ipl,x='city')
plt.xticks(rotation=90)
plt.xlabel("Cities",fontsize=16)
plt.ylabel("Number of matches played",fontsize=16)
plt.title("City with maximum matches played\n",fontsize=24)
```

```python
plt.figure(figsize=(20,6))
sns.countplot(data=ipl,x='toss_winner',palette='Set3')
plt.xticks(rotation=90)
plt.xlabel("Toss winners",fontsize=16)
plt.ylabel("Number of matches played",fontsize=16)
plt.title("team winning maximum tosses\n",fontsize=24)
```

```python
plt.figure(figsize=(20,6))
sns.countplot(data=ipl,x='player_of_match',palette='Set1')
plt.xticks(rotation=90)
plt.xlabel("Man of the Match",fontsize=16)
plt.ylabel("Number of matches played",fontsize=16)
plt.title("Man of the match winners\n",fontsize=24)
```
