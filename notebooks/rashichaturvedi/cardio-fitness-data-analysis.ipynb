```python
# Simple Linear Regression


#Load function from sklearn
from sklearn import linear_model

# Create linear regression object
regr = linear_model.LinearRegression()

y = df['Miles']
x = df[['Usage','Fitness']]

# Train the model using the training sets
regr.fit(x,y)
```
