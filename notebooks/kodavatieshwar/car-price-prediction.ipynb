## importing data

```python
cars_data=pd.read_csv('../input/vehicle-dataset-from-cardekho/CAR DETAILS FROM CAR DEKHO.csv')
```

## importing the dependencies

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
#checking the shape
cars_data.shape
```

```python
#getting some info of the data set
cars_data.info()
```

```python
# checking for number of missing values
cars_data.isnull().sum()
```

```python
#checking the distribution of categorical data
print(cars_data.fuel.value_counts())
print(cars_data.seller_type.value_counts())
print(cars_data.transmission.value_counts())
print(cars_data.owner.value_counts())
```

## encoding the categorical data

```python
# encoding fuel column
cars_data.replace({'fuel':{'Diesel':0,'Petrol':1,'CNG':2,'LPG':3,'Electric':4}},inplace=True)
# encoding transmission column
cars_data.replace({'transmission':{'Manual':0,'Automatic':1}},inplace=True)
# encoding seller_type column
cars_data.replace({'seller_type':{'Individual':0,'Dealer':1,'Trustmark Dealer':2}},inplace=True)
#encoding owner column
cars_data.replace({'owner':{'First Owner':0,'Second Owner':1,'Third Owner':2,'Fourth & Above Owner':3,'Test Drive Car':4}},inplace=True)
```

```python
cars_data.head()
```

## creating lables and spltting the data into training and testing data 

```python
X=cars_data.drop(['name','selling_price'],axis=1)
Y=cars_data['selling_price']
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2,random_state=2)
```

# Model Training

## Linear Regression

```python
# loading linear regression model
lin_reg_model=LinearRegression()
```

```python
lin_reg_model.fit(X_train,Y_train)
```

## model evaluation

```python
#predicting
predicted_values=lin_reg_model.predict(X_train)
```

```python
# R square error
error_score=metrics.r2_score(Y_train,predicted_values)
print("R square error:",error_score)
```

### visualize actual and predicted values

```python
plt.scatter(Y_train,predicted_values,c='r')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Price')
plt.show()
```

```python
prediction_values=lin_reg_model.predict(X_test)
```

```python
error_score1=metrics.r2_score(Y_test,prediction_values)
print("R square error:",error_score1)
```

```python
plt.scatter(Y_test,prediction_values,c='r')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Price')
plt.show()
```

## Lasso Regression

```python
lasso_reg_model=Lasso()
lasso_reg_model.fit(X_train,Y_train)
```

```python
predicted_values1=lasso_reg_model.predict(X_train)
```

```python
# R square error
error_score2=metrics.r2_score(Y_train,predicted_values1)
print("R square error:",error_score2)
```

```python
plt.scatter(Y_train,predicted_values1,c='g')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Price vs Predicted Price')
plt.show()
```

```python
prediction_values1=lasso_reg_model.predict(X_test)
```
