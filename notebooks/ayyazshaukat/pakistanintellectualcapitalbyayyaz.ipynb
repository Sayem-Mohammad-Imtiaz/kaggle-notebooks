```python
fileName = '/kaggle/input/pakistanintellectualcapitalcs/Pakistan Intellectual Capital - Computer Science - Ver 1.csv'
main_data_frame = pd.read_csv(fileName, encoding='unicode_escape')
```

```python
len(main_data_frame)
```

```python
#Data Cleaning
degrees = ["Bachelors","Master","BS","MS","MPhil","PhD","PostDoc"]

    
main_data_frame.loc[main_data_frame['Terminal Degree']=='Bachelors'] = 'BS'
main_data_frame.loc[main_data_frame['Terminal Degree']=='BA'] = 'BS'
main_data_frame.loc[main_data_frame['Terminal Degree']=='BSCS','Terminal Degree'] = 'BS'
main_data_frame.loc[main_data_frame['Terminal Degree']=='BCS'] = 'BS'
main_data_frame.loc[main_data_frame['Terminal Degree']=='BIT'] = 'BS'
main_data_frame.loc[main_data_frame['Terminal Degree']=='BSIT'] = 'BS'


main_data_frame.loc[main_data_frame['Terminal Degree']=='MSC',         'Terminal Degree'] = 'Master'
main_data_frame.loc[main_data_frame['Terminal Degree']=='M.S',         'Terminal Degree'] = 'MS'
main_data_frame.loc[main_data_frame['Terminal Degree']=='Master',      'Terminal Degree'] = 'MS'
main_data_frame.loc[main_data_frame['Terminal Degree']=='M.Sc',        'Terminal Degree'] = 'MS'
main_data_frame.loc[main_data_frame['Terminal Degree']=='MCS',         'Terminal Degree'] = 'MS'
main_data_frame.loc[main_data_frame['Terminal Degree']==' MCS',        'Terminal Degree'] = 'MS'
main_data_frame.loc[main_data_frame['Terminal Degree']=='MS (Scholar)','Terminal Degree'] = 'MS'
main_data_frame.loc[main_data_frame['Terminal Degree']=='MSCS','Terminal Degree'] = 'MS'

main_data_frame.loc[main_data_frame['Terminal Degree']=='Mphil',                 'Terminal Degree'] = 'MPhil'
main_data_frame.loc[main_data_frame['Terminal Degree']=='M.Phil',                'Terminal Degree'] = 'MPhil'
main_data_frame.loc[main_data_frame['Terminal Degree']=='M.Phil leading to PhD', 'Terminal Degree'] = 'MPhil'

main_data_frame.loc[main_data_frame['Terminal Degree']=='Phd',           'Terminal Degree'] = 'PhD'
main_data_frame.loc[main_data_frame['Terminal Degree']=='Ph.D(Scholar)', 'Terminal Degree'] = 'PhD'
main_data_frame.loc[main_data_frame['Terminal Degree']=='Ph.D (Scholar)','Terminal Degree'] = 'PhD'
main_data_frame.loc[main_data_frame['Terminal Degree']=='Ph.D',          'Terminal Degree'] = 'PhD'

main_data_frame.loc[main_data_frame['Terminal Degree']=='Doctor of Professional Studies','Terminal Degree'] = 'PhD'

main_data_frame.loc[main_data_frame['Terminal Degree']=='Post Doc', 'Terminal Degree'] = 'PostDoc'


#Cleaning Country column
main_data_frame['Country'] = main_data_frame['Country'].str.strip()
main_data_frame.loc[main_data_frame['Country'] == "Urbana", "Country"] = "USA"
main_data_frame.loc[main_data_frame['Country'] == "Macau", "Country"] = "China"

main_data_frame['Graduated from'] = main_data_frame['Graduated from'].str.strip()

# Cleaning Graduated from column
main_data_frame.loc[main_data_frame['Graduated from']=='Bahria University,Islamabad', 'Graduated from'] = 'Bahria University'
main_data_frame.loc[main_data_frame['Graduated from']=='Beijing Institute of Technology Beijing', 'Graduated from'] = 'Beijing Institute of Technology'

    
main_data_frame.loc[main_data_frame['Graduated from']=='COMSATS Institute of Information Technology', 'Graduated from'] = 'COMSATS Institute of Information Technology'
main_data_frame.loc[main_data_frame['Graduated from']=='COMSATS Institute of Information Technology,Islamabad', 'Graduated from'] = 'COMSATS Institute of Information Technology'
main_data_frame.loc[main_data_frame['Graduated from']=='COMSATS Institute of Information Technology,Lahore', 'Graduated from']= 'COMSATS Institute of Information Technology'
main_data_frame.loc[main_data_frame['Graduated from']=='COMSATS Institute of Information Technology,Vehari', 'Graduated from'] = 'COMSATS Institute of Information Technology'
main_data_frame.loc[main_data_frame['Graduated from']=='COMSATS Institute of Information Technology,Wah Cantt', 'Graduated from'] = 'COMSATS Institute of Information Technology'
```

```python
#Which area of interest/expertise is in abundance in Pakistan and where we need more people?
#finding 'Each Type of interest/expertise' and teacher's count in each 'interest' 
```

```python
interests_dictionary = { "dbms":0, "database":0,
                         "object oriented programming":0, "oop":0,
                         "quality assurance":0, "testing":0,
                         "artificial intelligence" : 0, "security" : 0,
                         "cloud":0,"cloud computing":0,

                         "software engineering":0, "telecommunication":0, "network" :0, "operating system":0, "multimedia":0, "mining":0,  "computation theory":0, 
                         "algorithms":0, "data science":0 , "enterpreneurship":0, "accounting":0, "wireless":0, "robotics": 0, "data structures": 0, "programming":0,
                         "big data":0,  "machine learning":0, "distributed computing":0, 'computer architecture': 0, 'software architecture': 0,
                          "graph":0, "project management" :0 ,"mathematics" :0, "security":0, "management":0
                        }
```

```python
for row in main_data_frame.itertuples():
    if( type(row[11]) is not float):
        emp_interest = row[11].lower()
        for interest in interests_dictionary.keys():
            if(interest in emp_interest):
                interests_dictionary[interest] += 1
            
interests_dictionary = {k: v for k, v in sorted(interests_dictionary.items(),reverse=True, key=lambda item: item[1])} 
 
interests_dictionary
```

```python
import seaborn as sns
import matplotlib.pyplot as plt 

most_interests = {k: v for k, v in interests_dictionary.items() if v > 20}

plt.figure(figsize = (20, 8))

sns.barplot(x = [k for k, v in most_interests.items()] , y = [v for k, v in most_interests.items()], color = "Blue", label = "count")


plt.xlabel("Interests", fontsize = 20)
plt.ylabel("Count", fontsize = 20)
plt.title("Teacher's Interests",fontsize = 30)
plt.xticks(rotation = 80, fontsize = 20)
plt.yticks(fontsize = 20)
plt.legend()
plt.show()
```

```python
least_interests = {k: v for k, v in interests_dictionary.items() if v <= 19}

plt.figure(figsize = (20, 8))

sns.barplot(x = [k for k, v in least_interests.items()] , y = [v for k, v in least_interests.items()], color = "Blue", label = "count")


plt.xlabel("Interests", fontsize = 20)
plt.ylabel("Count", fontsize = 20)
plt.title("Teacher's Interests",fontsize = 30)
plt.xticks(rotation = 80, fontsize = 20)
plt.yticks(fontsize = 20)
plt.legend()
plt.show()
```
