**In brief:** we conclude that no gene can be considered as "biomarker" i.e. good enough predictor for the disease, at least by the simple
analysis considered below. 

More details:

ALS ( Amyotrophic lateral sclerosis ) transciptomics data is considered - it contains data matrices X and targets y.
There are three pairs (X1,y1), (X2,y2), (X3,y3) are given in 3 files. (Actually data matries X1 and X2 are subsets of X3).


Data contains huge amount of features - 53859 genes expression, while we have 112, 87, 133 targets in each of the three cases,
so it is exremely easy to overfit.

Here we proceed with simple  one variable analysis - i.e. calculate predictive power of each feature (on its own, i.e. without any model) 
with respect to targets.
We consider roc_auc as measure of predictive power. 

### Conclusion:
We migth see that some features show not so bad predictive power. However most probably it is just illusion - we boostrap random data of the same sizes and see that out of 53859 random  data one might find similarly "good" predictors. 

Thus at the moment we cannot reliably conclude that any of genes has good enough predictive power. I.e. can serve as "biomarker".
Additional analysis is required.

# Bulbar vs limb data

```python
fn = '/kaggle/input/end-als/end-als/transcriptomics-data/DESeq2/bulbar_vs_limb.csv'
# fn = '/kaggle/input/end-als/end-als/transcriptomics-data/DESeq2/median_low_vs_high.csv'
#fn = '/kaggle/input/end-als/end-als/transcriptomics-data/DESeq2/ctrl_vs_case.csv'
df = pd.read_csv(fn)
df
```

```python
y = df.iloc[:,1]
X = df.iloc[:,2:].values

import time 
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt

s1 = []
t0 = time.time()
for i in range(X.shape[1]):
    u = X[:,i] 
    r = roc_auc_score(y, u )
    s1.append(r)
print(time.time()-t0,'seconds passed')
print('Statistic for genes expressions: ')
print( pd.Series(s1).describe())

s2 = []
t0 = time.time()
for i in range(X.shape[1]):
    u = np.random.randn(len(y))
    r = roc_auc_score(y, u )
    s2.append(r)
print(time.time()-t0,'seconds passed')
print('Statistic random data: ')
print( pd.Series(s2).describe())

plt.plot(np.sort(s1), label = 'genes' )
plt.plot(np.sort(s2), label = 'random' )
plt.legend()
plt.title('ordered values of rocauc')
plt.show()
```
