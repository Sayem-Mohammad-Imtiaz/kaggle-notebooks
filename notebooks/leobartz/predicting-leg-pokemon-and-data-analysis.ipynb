## 3.0 Cleaning the Database

There is only one column (Type 2) with missing values, I have decided to eliminate this Column in my analysis because cleaning NaN values will lose to much data and I think it is not good to predict an stats like Type, because thei are very useful on game balance.

I will eliminate the # Column too and set a new index Name.

The names of Pokemons contained extra and unneeded text, I will emoved all the text before "Mega".  

### 3.1 Check for null and missing values

```python
# Fill empty and NaNs values with NaN
pokemon = pokemon.fillna(np.nan)

# Check for Null values
pokemon.isnull().sum()
```

**Type 2** has a lot of missing Values, the other columns seems to be ok.

```python
# Infos

pokemon.info()
```

```python
# Dtypes

pokemon.dtypes
```

### 3.2 Cleaning Data

```python
# Drop "Type2" Column

pokemon = pokemon.drop(["Type 2"],axis=1) 
```

```python
pokemon.head(2)
```

```python
# Change the index to "Name" column

pokemon = pokemon.set_index("Name")
```

```python
# Drop "# "Column

pokemon = pokemon.drop(["#"],axis=1) 
```

```python
pokemon.head(2)
```

```python
# Removed all the text before "Mega"  

pokemon.index = pokemon.index.str.replace(".*(?=Mega)", "")

pokemon.head()
```

```python
# Summarie and statistics

pokemon.describe()
```

Now the Data is ready for analysis.

### 3.4 New DF without Legendary

```python
# Drop Legendaries


pokemon_nl = pokemon[~pokemon[pokemon["Legendary"]==False].isin(legendary)] # Drop Pokemons with True in Legendary

pokemon_nl = pokemon_nl.drop(["Legendary"],axis=1) # Removing Legendary column


                                             
pokemon_nl.head()
```

I'm gonna try to predict if a Pokemon is Legendary or not.

```python
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn import preprocessing
from sklearn import tree
from sklearn.metrics import f1_score
```

### Features and labels

```python
features = pokemon.drop("Legendary", axis =1)
labels = pokemon["Legendary"]
```

### Splitting Data into Train and Test

```python

from sklearn.cross_validation import train_test_split

features_train, features_test, labels_train, labels_test = train_test_split(features, 
                                                                            labels, test_size=0.30, 
                                                                            random_state = 42)
```

## Selecting best Algorithm

```python

from sklearn.metrics import recall_score
from sklearn.metrics import precision_score

from sklearn.naive_bayes import GaussianNB
clf_g = GaussianNB()
clf_g.fit(features_train, labels_train)

clf_g.predict(features_test)

print ("Naive Bayes")
print ("Accuracy: "), clf_g.score(features_test, labels_test)

y_pred = clf_g.predict(features_test)

print ("Precision: "), precision_score(labels_test, y_pred) 
print ("Recall: "), recall_score(labels_test, y_pred)

```

```python

### Deciosion Tree
from sklearn import tree
clf_t = tree.DecisionTreeClassifier(random_state=42)

clf_t.fit(features_train, labels_train)
print ("Decision Tree")
print ("Accuracy_t: "),clf_t.score(features_test, labels_test)

y_pred_t = clf_t.predict(features_test)

print ("Precision_t: "), precision_score(labels_test, y_pred_t) 
print ("Recall_t: "), recall_score(labels_test, y_pred_t)
```

```python
### K-nearest Neighbors
from sklearn.neighbors import KNeighborsClassifier
clf_KNN = KNeighborsClassifier()
clf_KNN.fit(features_train, labels_train) 

y_pred_KNN = clf_KNN.predict(features_test)

print ("K-Nearest Neighbors")
print ("Accuracy: ", clf_KNN.score(features_test, labels_test))

print ("Precision_KNN: ", precision_score(labels_test, y_pred_KNN)) 
print ("Recall_KNN: ", recall_score(labels_test, y_pred_KNN) )
```

```python
### SVM

from sklearn.svm import SVC
clf_SVM = SVC(kernel='linear')
clf_SVM.fit(features_train, labels_train) 

y_pred_SVM = clf_SVM.predict(features_test)

print ("SVM")
print ("Accuracy: ", clf_SVM.score(features_test, labels_test))

print ("Precision_SVM: ", precision_score(labels_test, y_pred_SVM)) 
print ("Recall_SVM: ", recall_score(labels_test, y_pred_SVM)) 
```

### Selecting best parameters

```python

kn = {
    "select__k" : range(1,9),
    "clf__n_neighbors" : [5, 15, 20],
    "clf__weights" : ["uniform","distance"],
    "clf__algorithm" : ["ball_tree", "kd_tree", "brute"],    
}




nb = {
        "select__k" : range(1,9)
}


dt = {
    "select__k" : range(1,9),
    "clf__min_samples_split" : [2, 5, 10, 20, 40],
    "clf__criterion" : ["gini", "entropy"],
    "clf__splitter" : ["best","random"],
    "clf__max_depth": [None, 1, 2, 5, 10],
    "clf__min_samples_leaf":  [1, 2, 3, 4, 5, 6, 7, 8],
    "clf__random_state": [42] # manter os testes comparaveis
}
            
def find_best(classifier, parameters):
    estimators = [("select",SelectKBest()),("clf",classifier)]  #("scaling", scaler),
    pipe = Pipeline(estimators)
    
    sss = StratifiedShuffleSplit(n_splits=3, test_size=0.3, random_state=42)
    sss.get_n_splits(features,labels)
    
    result = GridSearchCV(pipe, parameters, cv = sss, scoring="f1")
    result.fit(features, labels)
    clf = result.best_estimator_
    
    print (result.best_params_)
    
    clf = classifier
    clf.fit(features_train, labels_train)
    clf.predict(features_test)
    
    
    print ("Accuracy: ", clf.score(features_test, labels_test))

    y_pred = clf.predict(features_test)

    print ("Precision: ", precision_score(labels_test, y_pred)) 
    print ("Recall: ", recall_score(labels_test, y_pred))   
    print ("F1 Score: ", f1_score(labels_test, y_pred))
```
