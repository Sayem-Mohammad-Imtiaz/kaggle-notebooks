```python
data = pd.read_csv("/kaggle/input/pakistans-largest-ecommerce-dataset/Pakistan Largest Ecommerce Dataset.csv")
data.head()
```

```python
data.shape
```

```python
data = data.drop(['Unnamed: 21', 'Unnamed: 22', 'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25'], axis=1)
```

```python
data.isna().sum()
```

```python
data = data.dropna()
data.head()
```

# Visualization

```python
category_total = data.groupby(["category_name_1"]).sum()["grand_total"].reset_index()
category_total
```

```python
fig = px.bar(category_total, x="category_name_1", y="grand_total", color="category_name_1")

fig.show()
```

```python
complited = data[(data["status"] == 'complete') | (data["status"] == 'received')]
complited = complited[["category_name_1", "grand_total"]].groupby(["category_name_1"]).sum()["grand_total"].reset_index()
complited
```

```python
fig = px.bar(complited, x="category_name_1", y="grand_total", color="category_name_1")

fig.show()
```
