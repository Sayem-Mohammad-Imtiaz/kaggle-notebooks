```python
import scipy.stats as stats
```

```python

def compare_means(col, cat='target'):
    col_0 = df[col][df[cat] == 1]
    col_1 = df[col][df[cat] == 0]

    print(col)
    print(stats.f_oneway(col_0, col_1))
    print(f'{col} (0): ', col_0.mean())
    print(f'{col} (1): ',col_1.mean())
    print()
```

```python
compare_means('trestbps')


compare_means('chol')

print('maximum heart rate achieved')
compare_means('thalach')

compare_means('oldpeak')
```
