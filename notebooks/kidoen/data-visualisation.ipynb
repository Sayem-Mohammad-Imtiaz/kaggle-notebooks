```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
color = sns.color_palette()
%matplotlib inline
import plotly.offline as py
from plotly import tools
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.express as px
pd.set_option('max_columns', 100)

from bokeh.models import Panel, Tabs
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.layouts import gridplot
output_notebook()
```

```python
df = pd.read_csv("/kaggle/input/nifty-indices-dataset/NIFTY 100.csv",parse_dates=['Date'])
```

```python
def scatter_plot(count,color):
    trace = go.Scatter(
            x = count.index[::-1],
            y = count.values[::-1],
            showlegend=False,
            marker=dict(color=color))
    return trace
    
```

```python
layout = go.Layout(
    title=go.layout.Title(
        text="NIFTY 100 Closing values over time",
        x=0.5
    ),
    xaxis_title="Date",
    yaxis_title="Index close value",
    font=dict(size=14),
    width=800,
    height=500,
)
```

```python
count = df['Close']
index = df['Date']
trace = scatter_plot(count,'red')
data = [trace]
fig = go.Figure(data=data,layout=layout)
py.iplot(fig,filename='fig')
```

```python
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df_month = df.groupby(['Year','Month']).first().reset_index()
```
