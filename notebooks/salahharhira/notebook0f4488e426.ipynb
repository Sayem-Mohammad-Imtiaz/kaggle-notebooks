```python
train_data=pd.read_csv("/kaggle/input/titanic/train_data.csv")
train_data.head()
```

```python
test_data=pd.read_csv("/kaggle/input/titanic/test_data.csv")
test_data.head()
```

```python
age=train_data.loc[ train_data.Age > 0.4]['Survived']
```

```python
sum(age)/(len(age))
```

```python
women=train_data.loc[ train_data.Sex == 0]['Survived']
sum(women)/len(women)
```

```python
male=train_data.loc[ train_data.Sex == 1]['Survived']
sum(male)/len(male)
```

```python
target=train_data["Survived"]
```

```python
f=["Age","Sex","Pclass_1","Pclass_2","Pclass_3"]
train_data[f]
```

```python
x=pd.get_dummies(train_data[f])
x_test=pd.get_dummies(test_data[f])
```

```python
model=RandomForestClassifier(n_estimators=100)
model.fit(x,target)
pred=model.predict(x_test)
```

```python
out=pd.DataFrame({'PassengerId':test_data.PassengerId,'Survived':pred})
```
