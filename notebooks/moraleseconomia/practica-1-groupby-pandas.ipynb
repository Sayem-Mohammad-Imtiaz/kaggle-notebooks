```python
import pandas as pd

dtypes = {
    "first_name": "category",
    "gender": "category",
    "type": "category",
    "state": "category",
    "party": "category",
}
df = pd.read_csv(
    "../input/legislatorshistorial/legislators-historical.csv",
    dtype=dtypes,
    usecols=list(dtypes) + ["birthday", "last_name"],
    parse_dates=["birthday"]
)
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.tail()
```

```python
n_by_state = df.groupby("state")["last_name"].count() #Comentario, agrupa "state", cuenta "last_name"
```

```python
n_by_state.head(10)
```

```python
df.groupby("party")["state"].count().head(10) #Comentario, agrupa "party", cuenta "state"
```

```python
df.groupby(["state", "gender"])["last_name"].count() 
#Agrupa por "state" y "gender", cuenta los "last_names"
```

```python
n_by_state_gender = df.groupby(["state", "gender"])["last_name"].count()
```

```python
type(n_by_state_gender)
```

```python
n_by_state_gender.index[:5] #vemos que es MultiIndex
```

```python
df.groupby(["state", "gender"], as_index=False)["last_name"].count() 
# Una manera de quitar quizás los innecesarios índices
```

```python
df.groupby("state", sort=False)["last_name"].count().head(10)  #sort=False quita que se vayan a ordenar
```

```python
by_state = df.groupby("state")
print(by_state)
```

```python
for state, frame in by_state:
     print(f"First 2 entries for {state!r}")
     print("------------------------")
     print(frame.head(2), end="\n\n")
```

```python
by_state.groups["PA"]
```

```python
by_state.get_group("PA") 
#Conseguir una pieza de las agrupaciones, lo que pertenece a "state" pero después de groupby
#This is virtually equivalent to using .loc[]. You could get the same output with something like df.loc[df["state"] == "PA"].
```

```python
df.loc[df["state"] == "PA"] #Lo de arriba es similar a esto
```

```python
# Usando agg

df.groupby('state').agg({'gender':'count'}).head() #Lanza un dataframe index
```

```python
# No me gusta los index
df.groupby('state', as_index=False).agg({'gender':'count'}).head()
```

```python
# Usando aggregate, pero no hay cuenta de caracteres
df.groupby('state', as_index=False)['gender'].aggregate(np.sum).head()
```

```python
df.groupby('state', as_index=False).agg({'gender':pd.Series.unique}).head()
```

```python
df.groupby('state', as_index=False).agg({'gender':pd.Series.nunique}).head()
```

```python
# Personas sobrevivieron o no por género
df1.groupby('Sex')['Survived'].sum()
```

```python
df1.groupby('Sex').aggregate(np.sum)  #Puede sumar toda la columna
```
