```python
df = pd.read_csv('../input/videogamesales/vgsales.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,15))
sns.countplot(data=df, y="Platform")
```

```python
for i in df:
    print("%d values are missing in column %s"%(df[str(i)].isnull().sum(), str(i)))
```

```python
df = df.fillna(method="bfill")
```

```python
for i in df:
    print("%d values are missing in column %s"%(df[str(i)].isnull().sum(), str(i)))
```

```python
df["Year"] = pd.to_datetime(df["Year"], format="%Y")
```

```python
df["Year"]
```

```python
plt.figure(figsize=(30,20))
sns.lineplot(x="Year", y="JP_Sales", data = df, ci=None, label="JP Sales")
sns.lineplot(x="Year", y="EU_Sales", data = df, ci=None, label="Europe Sales")
sns.lineplot(x="Year", y="NA_Sales", data = df, ci=None, label="North America Sales")
sns.lineplot(x="Year", y="Other_Sales", data = df, ci=None, label="Other Sales")
sns.lineplot(x="Year", y="Global_Sales", data = df, ci=None, label="Global Sales")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title("Sales by Year",fontsize=24)
plt.xlabel("Year",fontsize=20)
plt.ylabel("Sales",fontsize=20)
plt.legend(prop={'size': 30})
```

```python
plt.figure(figsize=(100,150))
sns.countplot(data=df, y="Publisher")
plt.savefig('testplot.png')
```

```python
plt.figure(figsize=(30,20))
sns.lineplot(x="Year", y="Global_Sales", data = df, ci=None, hue="Genre", size="Genre")
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.title("Sales by Year",fontsize=24)
plt.xlabel("Year",fontsize=20)
plt.ylabel("Sales",fontsize=20)
plt.legend(prop={'size': 30})
```
