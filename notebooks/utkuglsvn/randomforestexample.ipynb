###  RANDOM FOREST ALGORİTMASI

Rastgele Orman, verilen veri kümesinin çeşitli alt kümelerinde bir dizi karar ağacı içeren ve bu veri kümesinin tahmini doğruluğunu artırmak için ortalamayı alan bir sınıflandırıcıdır.

Bir veri setinin belirli parçalara bölerek her parçadan yeni bir karar ağacı elde eder, daha sonra o ağaçları birleştirerek sınıflama ya da regresyon yapılmasını sağlayan modeldir. Denetimli(Supervised) sınıflandırma algoritmasıdır. Kullanıcıdan iki parametre alır; m(En iyi bölünmeyi seçmek için düğüm değişken sayısı), N(Geliştirilecek ağaç sayısı)

Ormandaki daha fazla ağaç daha yüksek doğruluk sağlar ve aşırı uyum sorununu önler.



### Random Forest Nasıl Çalışır

Random Forest algoritması nasıl çalışır?
Rastgele Orman ilk olarak iki aşamalı olarak çalışır, N karar ağacını birleştirerek rastgele orman oluşturmak, ikincisi ise ilk aşamada oluşturulan her ağaç için tahminler yapmaktır.

Çalışma süreci aşağıdaki adımlar ve şemada açıklanabilir:

Adım-1: Egzersiz setinden rastgele K veri noktaları seçin.

Adım-2: Seçilen veri noktalarıyla (Altkümeler) ilişkili karar ağaçlarını oluşturun.

Adım-3: Oluşturmak istediğiniz karar ağaçları için N sayısını seçin.

Adım-4: Adım 1 ve 2'yi tekrarlayın.

Adım-5: Yeni veri noktaları için, her karar ağacının tahminlerini bulun ve yeni veri noktalarını çoğunluk oylarını kazanan kategoriye atayın.


#### Avantaj ve Dezavantajları
* Hem regresyon hem de sınıflandırma işlemleri
* Çok kullanışlı ve kullanımı kolay bir algoritma
* Yeterli agac varsa ormanda ezberleme(over fitting) oranı azalır
* Tahmin oluşturmak yavaştır

#### Rastgele orman(Random Forest ve Karar agacı(Decision tree) arasındaki farklar
Rastgele Orman algoritması ile Karar Ağacı algoritması arasındaki fark, Rastgele Orman’da kök düğümü(Root Node) bulma ve düğümleri bölme işlemlerinin rastgele çalışıyor olmasıdır.


```python
data=pd.read_csv('/kaggle/input/default-of-credit-card-clients-dataset/UCI_Credit_Card.csv')
data.head()
#Bu veri setinde müşterilerin önceki davranışlara göre aylık parayı ödeyip ödemediklerini analiz edeceğiz
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=data.drop('default.payment.next.month',axis=1)
y=data['default.payment.next.month']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier()
rfc
```

```python
model=rfc.fit(X_train,y_train)
model
```

```python
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
print("accuracy:",accuracy_score(y_test,predictions))
print("accuracy:",precision_score(y_test,predictions))
print("accuracy:",recall_score(y_test,predictions))
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, predictions))
```

##### Hyperparamater:
n_estimators:Kaç adet karar ağacı
criterion:string,gini,entropy
max_features:ağacı bölerken kullanılacak özellik sayısı
max_depth:Ağaçların derinliği
min_samples_split:Min sample sayısı
min_samples_leaf: Dal(leaf) oluşturmak için sayısı

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

n_folds=5
paramaters={'max_depth': range(2,20,5)}
rfc=RandomForestClassifier()

rfc=GridSearchCV(rfc,paramaters,cv=n_folds,scoring="accuracy",return_train_score=True)
rfc.fit(X_train,y_train)
```

```python
scores=rfc.cv_results_
pd.DataFrame(scores).head()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.plot(scores["param_max_depth"],scores["mean_train_score"],label="training accuracy")
plt.plot(scores["param_max_depth"],scores["mean_test_score"],label="test_accuracy")
plt.xlabel("max_depth")
plt.ylabel("accuracy")
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV

n_folds=5

parameters={'n_estimators': range(100,700,400)}
rf=RandomForestClassifier(max_depth=4)
rf=GridSearchCV(rf,parameters,cv=n_folds,scoring="accuracy",return_train_score=True)
rf.fit(X_train,y_train)
```

```python
scores=rf.cv_results_
pd.DataFrame(scores).head()
```

```python
plt.figure()
plt.plot(scores["param_n_estimators"],scores["mean_train_score"],label="training accuracy")
plt.plot(scores["param_n_estimators"],scores["mean_test_score"],label="test_accuracy")
plt.xlabel("n estimatros")
plt.ylabel("accuracy")
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
param_grid = {
    'bootstrap': [True],
    'max_depth': [4, 8, 12],
    'max_features': [2, 4,8],
    'min_samples_leaf': [3, 5, 7],
    'min_samples_split': [8, 10, 12],
    'n_estimators': [50, 100, 200]
}
rf = RandomForestRegressor()
rfc = RandomForestRegressor()
grid_search = GridSearchCV(estimator = rf, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
grid_search2 = GridSearchCV(estimator = rfc, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)
```

```python
grid_search.fit(X_train,y_train)
```

```python
#en iyi parametre
print('Best paramters:',grid_search.best_params_)
pd.DataFrame(grid_search.cv_results_).head()
```

```python
grid_search2.fit(X_train,y_train)
```

```python
#en iyi parametre
print('Best paramters:',grid_search2.best_params_)
```
