```python
df = pd.read_csv('../input/us-accidents/US_Accidents_Dec19.csv')
df.head()
```

```python
df.isnull().sum()
```

As you can see above, there is no missing values for state and id which we are going to examine.

```python
df = df[['ID','State']]
df
```

```python
accidents_by_state = df.groupby('State').count().reset_index()
accidents_by_state
```

Let's order the counts in descending order and take top 10 states.

```python
sorted_values = accidents_by_state.sort_values(by='ID',ascending=False)
sorted_values.head()
```

```python
f, ax = plt.subplots(figsize=(15 ,15))
sns.barplot(x = sorted_values['State'], y = sorted_values['ID'])
plt.show()
```

As you can see in bar chart, California has highest number of car accidents.
