**Question :** Are girls doing better than boys in maths? 

```python
# let's visualize data distribution of girls vs boys for maths score
import seaborn as sns
sns.boxplot(x=df.gender,y=df['math score'],data=df)
# let's check if the dataset is balanced
# i.e to check if #girls appearing for maths exam is comparable to #boys
df.groupby(['gender'])['gender'].count()
```

What do we observe here?
* #Girls appearing for the test is comparable to #boys. So the data is balanced 
* The mean of the maths score for boys is higher than girls
    * Does a higher mean for boys indicates "for sure" that boys are doing better than girls? Is this finding statistically significant?

* In order to show if boys are indeed better in maths than girls rather this being a chance event, we need to do the below : 
    * Form a Hypothesis
    * From both the populations, take repeated samples of size = n
    * Calculate and save means of the samples each time
    * Now, the task is to apply statistical tests on the means obtained to calculate a test statistic and based on the set p-value accept or reject the stated hypothesis
   
* Let's frame the hypothesis now : 
    * Let the mean of the math score of girls(sample) = $\mu1$
    * Let the mean of the math score of boys(sample) = $\mu2$
    * The null hypothesis would be stated as : 
        $\mu1$ > $\mu2$
    * Hence, the alternate hypothesis would be stated as :
        $\mu1$ <= $\mu2$

```python
# let's separate out the girl's scores from the dataframe
girls = df[df['gender'] == 'female']['math score']
```

```python
# similar task for boys
boys = df[df['gender'] == 'male']['math score']
```

```python
# no need to execute below piece of code
'''
# we will do the below steps 50 times as we wish to draw 50 samples
# Step 1 : Draw 100 data-points (sample)
# Step 2 : Calculate mean of the sample
# Step 3 : Save it into a Series
from sklearn.utils import shuffle
import numpy as np
girls_means = []
boys_means = []
for i in range(1,50):
    girls = shuffle(girls)
    boys = shuffle(boys)
    mean_g = np.mean(girls[:100])
    mean_b = np.mean(boys[:100])
    girls_means.append(mean_g)
    boys_means.append(mean_b)

girls_means = pd.Series(girls_means)
boys_means = pd.Series(boys_means)
'''
```

* Now, since we are ready with two series both having 50 mean values created out of 50 samples drawn out of the population of girls and boys respectively,
let's perform a t-test to check if the hypothesis can be accepted or rejected

```python
from scipy import stats
stats.ttest_ind(girls,boys)
```

We observe here that the p-value is an extremely low value (much lower than the default 0.05, this would mean that there is significant evidence against the null-hypothesis. This means there is a data point found beyond the critical value which thus provides evidence that the null hypothesis which is mean of girls maths scores is greater than boys stands rejected

** Please note that the first value of the obtained result is the t-statistic, the second value is evident which is the p-value **
