```python
import pandas as pd
import numpy as np
```

### Read Dataset 

```python
netflix_data=pd.read_csv('../input/netflix-original-films-imdb-scores/NetflixOriginals.csv')
```

```python
netflix_data.head(2)
```

### Some Quick EDA

#### 1. Movies by Language

Results : Most of the movies are in English

```python

common_languages=netflix_data['Language'].value_counts().reset_index(name='total')
language_list=common_languages[common_languages['total']>=3]['index']
```

```python
common_languages[common_languages['total']>3].plot.bar(x='index', y='total', rot=90)
```

#### 2.Is there any seasonality for Movie Premiere

Result: There are some seasonality pattern around holidays but the sudden spike in 2020 accounts to the lockdown and the boom of the on demand video platforms resulting in more netflix originals 

```python
netflix_data['Premiere']= pd.to_datetime(netflix_data['Premiere'])
netflix_data['month_year'] = netflix_data['Premiere'].dt.to_period('M')
yeardata=netflix_data[netflix_data['month_year'] >= '2017-01']['month_year'].value_counts().reset_index(
name='totalcount')
yeardata=yeardata.sort_values('index', ascending=True)
yeardata.plot('index','totalcount')
```

#### 3. Whats the most popular Genre?

Result: The most popular Genre based on average IMDB rating are 
* Animation/Christmas/Comedy/Adventur
* Musical / Short
* Concert Film
* Anthology/Dark comedy


```python
netflix_data.groupby('Genre')['IMDB Score'].mean().reset_index(name='average_rating').\
sort_values('average_rating', ascending=False).head(4)
```

#### 4. Is there any correlation between Runtime and IMDB Score?

Result :
There is no correlation between Runtime and IMDB Score

```python
data = netflix_data[['Runtime','IMDB Score']]
data.plot('IMDB Score','Runtime')
correlation = data.corr(method='pearson')
correlation
```

#### 5. Whats the average run time of each Genre?

Result: It seems like Heist movies have an average runtime of 149 minutes and there are few genre that showed as low as 4 minutes which i imagine might be trailers/short clips

```python
netflix_data.groupby('Genre')['Runtime'].mean().reset_index(
name='average_runtime').sort_values('average_runtime', ascending=False)
```
