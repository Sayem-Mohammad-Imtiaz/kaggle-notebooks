#### Accessing Items

- Use the index operator `[]` to access individual elements in the list. 
- List index starts from 0
- Access items from the end of the list by using negative indices 
- Use the operator `[start:end]` to access a range of items.
- Alternatively, leave blank to mean from start / until end `:<stop at index>`, `<start at index>:`

```python
my_list = [1, 2, 3, 4]
print(my_list[1])
print(my_list[-1])
print(my_list[1:])
```

### For loops

- Go through (or iterate over) every item in the list
- Use the `for .. in` syntax, as in `for item in my_list`
-  Each time through the loop, the variable `item` takes on the value of the next object in the collection. In the first iteration `item` is assigned the value of the first element in the list. In second iteration `item` is assigned the value of the second item in the list. This process continues until there are no more elements in the list, and the for loop exits.



```python
list_multiplied = []
for item in [1, 2, 3]:
  list_multiplied.append(item * item)
list_multiplied
```

### List comprehensions

A more consise way to do this in python is by using a list comprehension. 

- The for loop comes at the end, and the item being collected comes at the beginning.
- We can also add an optional `if` condition at the end


```python
my_list = [1, 2, 3]
[item * item for item in my_list]
```

```python
### Enumerate

- To get the index of the item, as well as the value, we can use the `enumerate` function. 
- The `enumerate` function gives you back two loop variables: The count of the current iteration, and the value of the current iteration.
```

```python
for index, item in enumerate(my_list):
    print(f"At index {index} is item {item}.")
```

## Dictionaries

- Dictionary is a similar concept to a regular dictionary of `Word: Definition`. We look up a word to get its definition. This `Word: Definition` we refer to as key-value pairs.
- It enables you to quickly lookup values using a key.
- Each key in the dictionary must be unique.

### Creating a dictionary

```python
dict()
{}
best_sellers = {
    '10-Day Green Smoothie Cleanse': 'JJ Smith',
    '11/22/63: A Novel': 'Stephen King',
    '12 Rules for Life: An Antidote to Chaos': 'Jordan B. Peterson',
    '1984 (Signet Classics)': 'George Orwell',
    '5,000 Awesome Facts (About Everything!) (National Geographic Kids)': 'National Geographic Kids'
}
```
