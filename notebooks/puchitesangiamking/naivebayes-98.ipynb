```python
url = 'https://raw.githubusercontent.com/mohitgupta-omg/Kaggle-SMS-Spam-Collection-Dataset-/master/spam.csv'
df = pd.read_csv('/kaggle/input/sms-spam-collection-dataset/spam.csv', encoding='latin-1')

arraysdf = df.values

print(df.values)

documentHam = df['v2'].tolist()
documentv1 = df['v1'].tolist()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer()   

count = CountVectorizer(stop_words = 'english')     ## use stop words 

count.fit(documentHam)

count.vocabulary_      
```

```python
from sklearn.model_selection import train_test_split

doc_arraySpam = count.transform(documentHam).toarray()     

X_train_spam, X_test_spam, Y_train_spam, Y_test_spam = train_test_split(doc_arraySpam, documentv1, test_size=0.1)
```

```python
from sklearn.naive_bayes import BernoulliNB    

BNB = BernoulliNB()
BNB.fit(X_train_spam,Y_train_spam)

resultBNB = BNB.score(X_test_spam,Y_test_spam)

print("Accuracy BernoulliNB is %.2f%% " % (resultBNB*100))
```
