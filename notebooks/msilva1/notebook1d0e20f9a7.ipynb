# Ciclo de Modelagem

1. Análise Exploratória: entender os dados (estatísticas descritivas e gráficos)

2. Modelagem

    2.1. Limpeza dos Dados (sp, SP, Sampa, São paulo -> SP)
    
    2.2. Criação de Variáveis ou feature engineering: Balance / EstimatedSalary -> proporção do quanto a pessoa tem em caixa com a gente dado salário; Geography (vários países) -> Continent
    
    2.3. Desenvolvimento do modelo: teste de vários algoritmos -> avalia as métricas de performance e escolhe possíveis candidatos
    
    2.4. Tunning de Hiperparâmetros: faz pros candidados -> avalia as métricas de performance dos modelos e escolhe o campeão

3. Apresentar pra área de negócio: ganho que o modelo traz -> a partir do modelo M conseguiríamos reduzir em x% a taxa de churn se fosse implentado N meses atrás até hoje, e deixaria de perder Z MM de reais

4. Deploy
    4.1. Batch: cálculo a cada X tempo a partir de uma base de dados
    
    4.2. Real Time: Cria API e recebe requests para cálculos em tempo real

5. Monitoria de modelo em produção: avaliar saúde do modelo e dos dados
