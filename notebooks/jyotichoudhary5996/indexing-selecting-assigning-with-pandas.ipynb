**Index-based selection**
selecting data based on its numerical position in the data.
**iloc** follows this approach.

To select the first row of data in a DataFrame, use the below way. 

```python
student.iloc[0]
```

Both loc and iloc are row-first, column-second. To get a column with iloc, we can do the following:

```python
student.iloc[:,0]
```

The : operator, means "everything". When combined with other selectors, however, it can be used to indicate a range of values. For example, to select the school column from just the first, second, and third row, we would do:-

```python
student.iloc[:3,0]
```

Or, to select just the second and third entries, we would do:

```python
student.iloc[1:3,0]
```

It's also possible to pass a list  of both rows and columns we need:-

```python
student.iloc[[0,1,2,3,4,5,6,7,8,21,13,12],[0,2,4]]
```

Negative numbers can be used in selection.
This will start counting forwards from the end of the values.
So for example here are the last five elements of the dataset. we will have option of using
df.tail(5) or df.iloc[-5:]

```python
student.iloc[-5] #the 5th last row data
```

```python
student.iloc[-5:] #data of last 5 rows
```

**Label-based selection**

**loc** follows this approach.
For example, to get the first entry in student, we would now do the following:

```python
student.loc[0,'school']
```

**iloc** is conceptually simpler than loc because it ignores the dataset's indices. When we use iloc we treat the dataset like a big matrix (a list of lists), one that we have to index into by position. 
**loc**, uses the information in the indices to do its work. Since dataset usually has meaningful indices, it's usually easier to do things using loc instead. 

For example, here's one operation that's much easier using loc

```python
student.loc[:,['school','age','Mjob','Fjob']]
```

**Manipulating the index**
Label-based selection derives its power from the labels in the index.
The set_index() method can be used to do the job

```python
student.set_index("school")
```

**Conditional selection**

```python
student.loc[(student.school=="MS") & (student.Mjob=="other") & (student.age>=19)]
```

Pandas comes with a few built-in conditional selectors.

The first is **isin**. 
isin is lets you select data whose value "is in" a list of values. 
For example, here's how we can use it to select students whose mother works in service sector or is a teacher. 

```python
student.loc[student.Mjob.isin (["services","teacher"])]
```

```python
student.loc[(student.Fjob.isin(["health","at_home"])) & (student.Mjob.isin(["teacher"]))]
```
