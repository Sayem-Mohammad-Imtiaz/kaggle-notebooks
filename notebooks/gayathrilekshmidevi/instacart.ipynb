```python
def split_dataframe(df, chunk_size=30000): 
    chunks = list()
    num_chunks = len(df) // chunk_size + 1
    for i in range(num_chunks):
        chunks.append(df[i*chunk_size:(i+1)*chunk_size])
    return chunks
```

```python
group_df_split = split_dataframe(group_df, 50)
```

```python

basket = (group_df
          .groupby(['product_id', 'product_name'])['add_to_cart_order']
          .sum().unstack().reset_index().fillna(0)
          .set_index('product_id'))
```

```python
basket
```

```python
#converting all positive vaues to 1 and everything else to 0
def my_encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets = basket.applymap(my_encode_units)
```

```python
basket_sets
```

```python
#Generatig frequent itemsets
frequent_itemsets = apriori(basket_sets, min_support=0.008, use_colnames=True)
```

```python
#generating rules
rules = association_rules(frequent_itemsets, metric="lift")
rules.sort_values('confidence', ascending = False, inplace = True)
rules.head(10)
```
