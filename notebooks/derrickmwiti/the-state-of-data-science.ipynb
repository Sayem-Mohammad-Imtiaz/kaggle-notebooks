```python
sal =mc[mc['CompensationAmount'].notnull()]
```

```python
def salary(x):
    x = x.replace(',', '')
    try:
        return float(x)
    except:
        return np.nan
    
```

```python
sal['CompensationAmount'] = sal['CompensationAmount'].apply(salary)
```

```python
usa = sal[sal['Country']=='United States']
```
