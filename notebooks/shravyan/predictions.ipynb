```python
from sklearn.model_selection import train_test_split
X= df.drop(["fetal_health"],axis=1)
Y= df["fetal_health"]
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
X=std.fit_transform(X)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=1)
```

**MLP Classifier**

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(250,250,250),max_iter=500,random_state=1)
mlp.fit(X_train,Y_train)
pred = mlp.predict(X_test)
predt=mlp.predict(X_train)
                   
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print("Test Dataset results")
print(classification_report(Y_test,pred))
print("Train dataset results")
print(classification_report(Y_train,predt))
```

```python
sns.heatmap(confusion_matrix(Y_test,pred),annot=True).set_title("Confusion Matrix MLP Classifier")
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=1,n_estimators=500,max_features=9,min_samples_leaf=5,max_depth=10)
rf.fit(X_train,Y_train)
predrf = rf.predict(X_test)
predrft = rf.predict(X_train)
```

```python
print("Test Dataset Results")
print(classification_report(Y_test,predrf))
print("Train Dataset Results")
print(classification_report(Y_train,predrft))

```

```python
sns.heatmap(confusion_matrix(Y_test,predrf),annot=True).set_title("Confusion Matrix Random Forest")
```

**Conclusion:**
MLP Classifier and Random Forest have almost same accuracy. However, prediction of Class 3 of Random Forest is better than MLP Classifier.
