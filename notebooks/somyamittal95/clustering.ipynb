```python
df1=df.loc[:,'Annual Income (k$)':'Spending Score (1-100)']
```

```python
import seaborn as sns
sns.pairplot(df1,diag_kind='kde')
```

We need to see which peaks are predominant in kde so that we can get an idea of optimum k value.
About 3 clusters can be seen in kde of spending score.

```python
from scipy.stats import zscore
df1=df1.apply(zscore)
```

```python
Model=KMeans(n_clusters=3)
Model
```

```python
df1=pd.DataFrame(df1)
Model.fit(df1)
Model.inertia_
```

```python
cluster_error=[]
for i in range(1,15):
    Model=KMeans(n_clusters=i,n_init=10)
    Model=Model.fit(df1)
    labels=Model.labels_
    centroids=Model.cluster_centers_
    cluster_error.append(Model.inertia_)
clusters_df=pd.DataFrame({'n_cluster':range(1,15),'error':cluster_error})
clusters_df
    
    
```

```python
#Knee Plot 
import matplotlib.pyplot as plt
plt.plot(clusters_df.n_cluster,clusters_df.error,marker='o')
```

```python
model=KMeans(n_clusters=6,n_init=15)
model.fit(df1)
centroids=model.cluster_centers_
```

```python
x=pd.DataFrame(centroids,columns=list(df1))
x
```

The  elbow plot we saw k=6 is optimum.So 6 clusters are formed. Thus, we conclude that category 3 has good annual income and still spendind score is really low. So we need to focus on this category alot.
