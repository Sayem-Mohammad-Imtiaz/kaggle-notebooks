### Data Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
train_df['gender']=le.fit_transform(train_df['gender'])
train_df['target']=le.fit_transform(train_df['target'])
train_df['relevent_experience']=le.fit_transform(train_df['relevent_experience'])
```

```python
train_df['company_size']=train_df['company_size'].replace(to_replace='50-99',value=75)
train_df['company_size']=train_df['company_size'].replace(to_replace='100-500',value=300)
train_df['company_size']=train_df['company_size'].replace(to_replace='10000+',value=10000)
train_df['company_size']=train_df['company_size'].replace(to_replace='10/49',value=30)
train_df['company_size']=train_df['company_size'].replace(to_replace='1000-4999',value=3000)
train_df['company_size']=train_df['company_size'].replace(to_replace='500-999',value=750)
train_df['company_size']=train_df['company_size'].replace(to_replace='5000-9999',value=7500)
train_df['company_size']=train_df['company_size'].replace(to_replace='<10',value=9)
```

Changing last new job values from string to int

```python
train_df['last_new_job']=train_df['last_new_job'].replace(to_replace='1',value=1)
train_df['last_new_job']=train_df['last_new_job'].replace(to_replace='>4',value=6)
train_df['last_new_job']=train_df['last_new_job'].replace(to_replace='2',value=2)
train_df['last_new_job']=train_df['last_new_job'].replace(to_replace='never',value=0)
train_df['last_new_job']=train_df['last_new_job'].replace(to_replace='4',value=4)

train_df['last_new_job']=train_df['last_new_job'].replace(to_replace='3',value=3)
```

```python
train_df=train_df.drop(['major_discipline'],axis=1)
```

# Splitting the Dataset

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
print(x_train.shape,y_train.shape)
```

# Balancing Dataset with SMOTE

```python
from imblearn.over_sampling import SMOTE
smt=SMOTE(random_state=0)
```

```python
x_train_smote,y_train_smote=SMOTE().fit_resample(x_train,y_train)
```

```python
print(x_train_smote.shape,y_train_smote.shape)
```

```python
from collections import Counter
print("Before SMOTE", Counter(y_train))
print("After SMOTE", Counter(y_train_smote))
```

# Using Machine Learning Algorithms

1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr_pred=lr.predict(x_test)
lr_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm_lr=confusion_matrix(y_test,lr_pred)
cm_lr
```

```python
acc_lr=accuracy_score(y_test,lr_pred)
acc_lr
```

2. Random Forest Classifier 

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=400)
rfc.fit(x_train,y_train)
```

```python
rfc_pred=rfc.predict(x_test)
rfc_pred
```

```python
cm_rfc=confusion_matrix(y_test,rfc_pred)
cm_rfc
```

```python
acc_rfc=accuracy_score(y_test,rfc_pred)
acc_rfc
```

3. Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dsc=DecisionTreeClassifier()
dsc.fit(x_train,y_train)
```

```python
dsc_pred=dsc.predict(x_test)
dsc_pred
```

```python
cm_dsc=confusion_matrix(y_test,dsc_pred)
cm_dsc
```

```python
acc_dsc=accuracy_score(y_test,dsc_pred)
acc_dsc
```

4. XGB Classifier 

```python
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(x_train,y_train)
```

```python
xgb_pred=xgb.predict(x_test)
xgb_pred
```

```python
xgb_dsc=confusion_matrix(y_test,xgb_pred)
xgb_dsc
```

```python
acc_xgb=accuracy_score(y_test,xgb_pred)
acc_xgb
```

# Neural Network

```python
from keras.models import Input,Model
from keras.layers import Dense
```

```python
i=Input(shape=[21])

layer1=Dense(units=21,activation='relu')(i)
layer2=Dense(units=12,activation='relu')(layer1)
layer3=Dense(units=8,activation='relu')(layer2)
out=Dense(units=1,activation='sigmoid')(layer3)

ann=Model(inputs=i,outputs=out)
```

```python
ann.summary()
```

# Flow-Chart

```python
from keras.utils.vis_utils import plot_model
```

```python
plot_model(ann, to_file='ann_model_plot.png', show_shapes=True, show_layer_names=True)
```

# Compiling the model

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

# Training the model

```python
ann.fit(x_train,y_train,batch_size=32,epochs=50)
```

```python
y_pred=ann.predict(x_test)
y_pred=(y_pred>0.5)
y_pred
```

```python
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
score=accuracy_score(y_test,y_pred)
score
```
