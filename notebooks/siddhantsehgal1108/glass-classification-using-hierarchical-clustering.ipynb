```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('/kaggle/input/glass/glass.csv')
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset['Type'].unique()
```

```python
x=dataset.iloc[:,:-1].values
```

```python
x
```

# Determining optimal number of clusters

```python
import scipy.cluster.hierarchy as sch 
dendrogram=sch.dendrogram(sch.linkage(x,method='ward'))
plt.title('Optimal number of clusters')
plt.xlabel('Glass composition')
plt.ylabel('Euclidean distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
y_pred=hc.fit_predict(x)
y_pred
```

# Visualisation for clusters

```python
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],s=100,c='red',label='Type 1')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],s=100,c='blue',label='Type 2')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],s=100,c='yellow',label='Type 3')

plt.title('Glass Classification')
plt.xlabel('Glass compostion')
plt.ylabel('RI')
plt.legend()
plt.show()
```

```python
from sklearn.metrics.cluster import silhouette_score
sc=silhouette_score(x,y_pred)
print(sc)
```
