```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv(os.path.join(dirname, filename))

#Solo se usan doláres ya que los datos para otras monedas son muy pocos datos.
data = data[(data.Currency == 'USD')]

data = data.drop(columns=['App Name', 'App Id', 'Currency', 'Developer Id', 'Free', 'Released','Developer Email', 'Developer Website', 'Privacy Policy', 'In App Purchases', 'Editors Choice', 'Installs', 'Ad Supported'])

#Variables numéricas:
ncal_labels = ['Rating', 'Rating Count', 'Minimum Installs', 'Maximum Installs', 'Price', 'Last Updated']
```

```python
data.head()
```

Correlation coefficient for numerical variables:

```python
corr_matrix = data.corr().to_numpy()
data.corr()
```

```python
fig1 = px.imshow(corr_matrix, x=ncal_labels, y=ncal_labels)
fig1.show()
```

```python
fig2 = px.scatter(data, x='Category', y='Rating', size='Maximum Installs', color='Category', range_y=[1.8,5])
fig2.show()
```

```python
fig3 = px.pie(data, values='Rating', names='Category')
fig3.show()
```

```python
data['Rating']= data['Rating'].astype(float)

fig4 = px.violin(data['Rating'], box=True, y='Rating')
fig4.show()
```

```python
fig5 = px.histogram(data, x='Rating', marginal="violin")
fig5.show()
```

```python
fig6 = px.scatter(data, y='Rating Count', x=['Minimum Installs','Maximum Installs'])
fig6.show()
```

```python
fig7 = px.scatter(data, y='Rating', x='Price')
fig7.show()
```

```python
fig8 = px.box(data, x='Category', y='Rating')
fig8.show()
```
