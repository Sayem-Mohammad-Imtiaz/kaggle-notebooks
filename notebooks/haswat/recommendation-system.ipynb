# **Recommendation System**

A recommendation system is a system that refers to predicting a number of items or data for future users, then making it a recommendation for the top item. One of the reasons why it is necessary to use a recommendation system is because users have many choices to use because of the prevalence of the internet.Recommeder system creates a similarity between the user and items and exploits the similarity between user/item to make recommendations.

```python
#Analysis of user rating given by the name 
no_of_rated_books_per_name = df.groupby(by='Name')['User Rating'].count().sort_values(ascending=False)
no_of_rated_books_per_name.head(5)
```

```python
#Analysis of user rating given by the Author 
no_of_rated_books_per_Author = df.groupby(by='Author')['User Rating'].count().sort_values(ascending=False)
no_of_rated_books_per_Author.head(5)
```

```python
C=df['User Rating'].mean()
m = df['Reviews'].quantile(0.1565)
q_books = df.copy().loc[df['Reviews'] >= m]
q_books.shape
print(m)
```

```python
def weighted_rating(df, m=m, C=C):
    v = df['Reviews']
    R = df['User Rating']
    # Perhitungan berdasarkan rumus IMDB
    a=(v/(v+m) * R) + (m/(m+v) * C)
    return a
```

```python
q_books['score'] = q_books.apply(weighted_rating, axis=1)
```

```python
#Top 10 rated books
q_books.groupby(['Name','Author','Genre'], as_index=False)[['User Rating', 'Reviews', 'score']].mean().sort_values(by='score', ascending=False).head(10)
```
