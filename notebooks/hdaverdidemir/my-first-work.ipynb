```python
#correlation map eğer aralarında +1 değeri varsa doğru orantılır.-1 değeri varsa ters orantılır. 0 değeri ise ilişkileri yoktur demektir.
#figsize görünen tablonun büyüklüğü,yazmazsak default olarak gelir.
f,ax=plt.subplots(figsize=(18,18))
sns.heatmap(data.corr(),annot=True, linewidths=.5,fmt='.1f', ax=ax)
#annot sayıların görünüp görünmemesi,linewidths kareler arası çizgi kalınlığı, fmt .1 noktadan sonra 1 basamak gösterir.
plt.show()
```

Matplotlib datayı görselleştirmeye yarar.
Line plot eğer x ekseni zamanda daha iyidir.
Scatter ise iki variable arasında ilişki (correlation) olup olmadığını göstermek için kullanılır.
Histogram datanın sıklığını yani frekansını göstermek içib kullanırız.

```python
#alpha saydamlık x eksesi pokemonlar y ekseni hız ve defans özellikleri
data.Speed.plot(kind='line', color='g',label='Speed', linewidth=1, alpha=0.5,grid=True, linestyle=':')
data.Defense.plot(color='r', label='Defense', linewidth=1, alpha=0.5,grid=True, linestyle='-.')
plt.legend(loc='upper right')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Line Plot')
plt.show()
```

```python
data.plot(kind='scatter', x='Attack', y='Defense',alpha = 0.5,color = 'red')
plt.xlabel('Attack')              # label = name of label
plt.ylabel('Defence')
plt.title('Attack Defense Scatter Plot')   
```

```python
# Histogram
# bins = number of bar in figure
data.Speed.plot(kind = 'hist',bins = 50,figsize = (12,12))
plt.show()
```

```python
# clf() = cleans it up again you can start a fresh
data.Speed.plot(kind = 'hist',bins = 50)
plt.clf()
# We cannot see plot due to clf()
```
