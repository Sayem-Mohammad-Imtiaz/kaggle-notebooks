```python
data.AveragePrice.plot(kind = 'hist',bins = 50,figsize = (10,10))
plt.xlabel("Average Prices")
plt.ylabel("Frequency")
plt.show()
```

```python
data_2015=data[data.year==2015]
data_2016=data[data.year==2016]
data_2017=data[data.year==2017]
data_2018=data[data.year==2018]
```

```python
plt.figure(figsize=(12,12))
plt.bar("2015",data_2015.AveragePrice.mean(),color="red")
plt.bar("2016",data_2016.AveragePrice.mean(),color="blue")
plt.bar("2017",data_2017.AveragePrice.mean(),color="green")
plt.bar("2018",data_2018.AveragePrice.mean(),color="orange")
plt.ylim(0,1.6)
plt.xlabel("Years")
plt.ylabel("Average Prices")
plt.grid()
plt.show()
```

```python
#Detecting prices which are higher than average prices in their year
price_2015=[i for i in data_2015.AveragePrice if i > data_2015.AveragePrice.mean()]
price_2016=[i for i in data_2016.AveragePrice if i > data_2016.AveragePrice.mean()]
price_2017=[i for i in data_2017.AveragePrice if i > data_2017.AveragePrice.mean()]
price_2018=[i for i in data_2018.AveragePrice if i > data_2018.AveragePrice.mean()]
print(len(price_2015))
print(len(price_2016))
print(len(price_2017))
print(len(price_2018))
```

```python
#Comparing 2 type of Avocado
plt.figure(figsize=(10,10))
plt.bar("Conventional",data.AveragePrice[data.type=="conventional"].mean())
plt.bar("Organic",data.AveragePrice[data.type=="organic"].mean())
plt.xlabel("Type of Avocado")
plt.ylabel("Average Price")
plt.ylim(0,1.8)
plt.grid()
plt.show()
```

```python
import warnings
warnings.filterwarnings("ignore")
data_list=[i for i in data.Date]
datetime_object = pd.to_datetime(data_list)
data["Date"]=datetime_object
```

```python
data2=data
data2=data2.set_index("Date")
data3=data2.resample("M").mean()
data3.head()
```

```python
data3.index[0].month_name()
plt.figure(figsize=(13,12))
plt.bar(data3.index.month_name(),data3.AveragePrice)
plt.xlabel("Months")
plt.ylabel("Average Price ")
plt.ylim(0,2)
plt.grid()
plt.show()
```

```python
#Sorting Average Prices by states
datasorted=data.groupby("State").mean()
datasorted=datasorted.sort_values("AveragePrice",ascending=False)
```

```python
#visualization
plt.figure(figsize=(17,12))
sns.barplot(x=datasorted.index,y=datasorted.AveragePrice)
plt.xticks(rotation= 75)
plt.xlabel('States')
plt.ylim(0,1.8)
plt.ylabel('Average Prices')
plt.title('Average Prices by States')
```
