```python
# Distribution of likes

# take the log of each column here
df_youtube['likes_log'] = np.log(df_youtube['likes'] + 1)
df_youtube['views_log'] = np.log(df_youtube['views'] + 1)
df_youtube['dislikes_log'] = np.log(df_youtube['dislikes'] + 1)
df_youtube['comment_log'] = np.log(df_youtube['comment_count'] + 1)

plt.figure(figsize = (12, 6))

plt.subplot(221)
g1 = sns.distplot(df_youtube['views_log'])
g1.set_title('Views log distribution')

plt.subplot(224)
g2 = sns.distplot(df_youtube['likes_log'], color='g')
g2.set_title('Likes log distribution')

plt.subplot(223)
g3 = sns.distplot(df_youtube['dislikes_log'], color="r")
g3.set_title('Dislikes log distribution')

plt.subplot(222)
g4 = sns.distplot(df_youtube['comment_log'])
g4.set_title('Comment log distribution')

plt.subplots_adjust(wspace = 0.2, hspace = 0.4,top = 0.9)
plt.show()
```

```python
# Look at the views, likes, and dislikes by category

df_youtube['category_name'] = np.nan

# read in the title from youtube categories to populate the category name
category_names = youtube_categories['items'].map(lambda row: row['snippet']['title'])

for i in range(0, len(category_names)):
    df_youtube.loc[(df_youtube['category_id'] == i+1), "category_name"] = category_names[i]
df_youtube['category_name']
```

```python
# graph the categories on a bar graph
plt.figure(figsize=(12,6))

category_counts = df_youtube.value_counts('category_name')

plt.title("Number of videos by category")
plt.xticks(rotation=90)

sns.barplot(x=category_counts.index, y=category_counts.values)
```

```python
import datetime

df_youtube["publish_hour"] = df_youtube.publish_time.map(lambda t: datetime.datetime.strptime(t, '%Y-%m-%dT%H:%M:%S.%fZ').hour)
df_youtube["publish_day"] = df_youtube.publish_time.map(lambda t: datetime.datetime.strptime(t, '%Y-%m-%dT%H:%M:%S.%fZ').day)
df_youtube["publish_weekday"] = df_youtube.publish_time.map(lambda t: datetime.datetime.strptime(t, '%Y-%m-%dT%H:%M:%S.%fZ').weekday())
df_youtube["publish_month"] = df_youtube.publish_time.map(lambda t: datetime.datetime.strptime(t, '%Y-%m-%dT%H:%M:%S.%fZ').month)
df_youtube["publish_year"] = df_youtube.publish_time.map(lambda t: datetime.datetime.strptime(t, '%Y-%m-%dT%H:%M:%S.%fZ').year)

df_youtube
```

```python
# plot the statistics for the most commonly posted times

def time_conversion(millitary_time):    
    if millitary_time > 11:
        return str(millitary_time-12)+"pm"
    elif millitary_time > 0:
        return str(int(millitary_time))+"am"
    return str(12)+"am"

most_commonly_posted_hour = df_youtube.groupby('publish_hour').publish_hour.count()
most_commonly_posted_hour.head()
# most_commonly_posted_hour.index = most_commonly_posted_hour.index.map(lambda t: time_conversion(t))
most_commonly_posted_hour

plt.figure(figsize = (12, 6))

sns.lineplot(data=most_commonly_posted_hour)
```

```python
plt.figure(figsize = (12,18))

plt.subplot(311)
df_youtube['like_rate'] =  df_youtube['likes'] / df_youtube['views'] * 100

g= sns.boxplot(x='category_name',y='like_rate',  data=df_youtube)
g.set_xticklabels(g.get_xticklabels(),rotation=45)
g.set_title("LIKE RATE DISTRIBUITIONS", fontsize=15)
g.set_xlabel("", fontsize=12)
g.set_ylabel("Like rate", fontsize=12)
```

```python
df_youtube['count_word']=df_youtube["title"].apply(lambda x: len(str(x).split()))
df_youtube['count_word_tags']=df_youtube["tags"].apply(lambda x: len(str(x).split()))


df_youtube['count_unique_word']=df_youtube["title"].apply(lambda x: len(set(str(x).split())))
df_youtube['count_unique_word_tags']=df_youtube["tags"].apply(lambda x: len(set(str(x).split())))

plt.figure(figsize = (12,18))

plt.subplot(421)
g1 = sns.distplot(df_youtube['count_word'], 
                  hist=False, label='Text')
g1 = sns.distplot(df_youtube['count_word_tags'], 
                  hist=False, label='Tags')
g1.set_title("COUNT WORDS DISTRIBUITION", fontsize=16)

plt.subplot(422)
g2 = sns.distplot(df_youtube['count_unique_word'],
                  hist=False, label='Text')
g2 = sns.distplot(df_youtube['count_unique_word_tags'], 
                  hist=False, label='Tags')
g2.set_title("COUNT UNIQUE DISTRIBUITION", fontsize=16)
```
