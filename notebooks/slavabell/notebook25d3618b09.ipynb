# Описание данных

<b>Атрибуты Трафика: </b>

ID: это уникальный идентификатор записи об аварии.

Source: указывает источник отчета об аварии (т. е. API, который сообщил об аварии.).

TMC: дорожно-транспортное происшествие может иметь код канала дорожных сообщений (TMC), который обеспечивает более подробное описание события.

Severity: показывает серьезность аварии, число от 1 до 4, где 1 указывает на наименьшее влияние на трафик (т. е. короткую задержку в результате аварии) и 4 указывает на значительное влияние на трафик (т. е. длительную задержку).

Start_Time: показывает время начала аварии в местном часовом поясе.
End_Time: показывает время окончания аварии в местном часовом поясе.

Start_Lat: показывает широту в координате GPS начальной точки.

Start_Lng: показывает долготу в координате GPS начальной точки.

End_Lat: показывает широту в координате GPS конечной точки.

End_Lng: показывает долготу в координате GPS конечной точки.

Расстояние (mi): протяженность дороги, на которую повлияла авария.

Описание: показывает описание аварии на естественном языке.

<b>Атрибуты Адреса: </b>

Number: показывает номер улицы в адресном поле.

Street: показывает название улицы в поле адреса.

Side: показывает относительную сторону улицы (справа/слева) в адресном поле.

City: показывает город в адресном поле.

State: показывает округ в адресном поле.

County: показывает состояние в адресном поле.

Zipcode: показывает zipcode в адресном поле.

Country: показывает страну в адресном поле.

Timezone: показывает часовой пояс в зависимости от места аварии (Восточный, Центральный и т. д.).

<b>Погодные Атрибуты </b>:

Airport_Code: обозначает метеорологическую станцию аэропорта, которая находится ближе всего к месту аварии.

Weather_Timestamp: показывает отметку времени записи наблюдения погоды (в местном времени).

Temperature(F) : показывает температуру (в градусах Фаренгейта).

Wind_Chill (F): показывает холод ветра (в градусах Фаренгейта).

Humidity(%): показывает влажность (в процентах).

Pressure(in): показывает давление воздуха (в дюймах).

Visibility(mi): показывает видимость (в милях).

Wind_Direction: показывает направление ветра.

Wind_Speed (mph): показывает скорость ветра (в милях в час).

Precipitation(in): показывает количество осадков в дюймах, если таковые имеются.

Weather_Condition: показывает погодные условия (дождь, снег, гроза, туман и т. д.).

<b>Атрибуты POI: </b>

Amenity: аннотация точки интереса (POI), которая указывает на наличие удобства в близлежащем месте.

Bump: Примечание POI, которое указывает на наличие лежачего полицейского или горба в близлежащем месте.

Crossing: Примечание POI, которое указывает на наличие пересечения в соседнем месте.

Give_Way: POI-аннотация, указывающая на наличие знака give_way в соседнем месте.

Junction: POI-аннотация, указывающая на наличие соединения в соседнем местоположении.

No_Exit: Примечание POI, которое указывает на наличие знака no_exit в соседнем местоположении.

Railway: Примечание POI, которое указывает на наличие железной дороги в близлежащем месте.

Roundabout: POI-аннотация, указывающая на наличие roundabout в соседнем месте.

Station: Примечание POI, которое указывает на наличие станции (автобус, поезд и т. д.) в соседнем месте.

Stop: Примечание POI, которое указывает на наличие знака остановки в соседнем месте.

Traffic_Calming: POI-аннотация, указывающая на наличие средств traffic_calming в близлежащем местоположении.

Traffic_Signal: аннотация POI, которая указывает на присутствие traffic_signal в соседнем местоположении.

Turning_Loop: POI-аннотация, указывающая на наличие turning_loop в соседнем местоположении.

<b>Период-дня:</b>

Sunrise_Sunset: показывает период дня (т. е. день или ночь) на основе восхода/захода солнца.

Civil_Twilight: показывает период суток (то есть день или ночь) на основе гражданских сумерек.

Nautical_Twilight: показывает период суток (то есть день или ночь) на основе морских сумерек.

Astronomical_Twilight: показывает период суток (то есть день или ночь) на основе астрономических сумерек.

```python
def missing_percentage(df): 
    total = df.isnull().sum().sort_values(ascending = False)[df.isnull().sum().sort_values(ascending = False) != 0]
    percent = round(df.isnull().sum().sort_values(ascending = False)/len(df)*100,2)[round(df.isnull().sum().sort_values(ascending = False)/len(df)*100,2) != 0]
    return pd.concat([total, percent], axis=1, keys=['Total','Percent'])

missing_percentage(df1)
```

Преобразуем сначала все bool в числовые данные с помощью функции

```python
def factorize (dataset):
    result = dataset.copy() 
    for cols in result.columns:
        if result.dtypes[cols] == np.bool:
            result[cols] = pd.factorize(result[cols],sort=True)[0]
    return result
df1 = factorize(df1)
df1.info()
```

```python
from sklearn import preprocessing

def into_integer (dataset):
    result = dataset.copy() #делаем копию
    encoders = {}
    for res in result.columns:  #пройдемся по колонкам
        if result.dtypes[res] == np.object:  #если тип колонки объект
            encoders[res] = preprocessing.LabelEncoder()  #сздаем кодировщик
            result[res] = encoders[res].fit_transform (result[res]) #применяем кодировщик к столбцу и перезаписываем столбец
    return result,encoders
df,encoders = into_integer(df1)
df.head()
```

### PCA
Cumulative Variance Ratio (кумулятивный размер объяснённой дисперсии данных) и Explained Variance Ratio (вклад каждой компоненты в общий объём объяснённой дисперсии)

```python
from sklearn.decomposition import PCA
pca_test = PCA(n_components=20)
pca_test.fit(X_train_augm)
sns.set(style='whitegrid')
plt.plot(np.cumsum(pca_test.explained_variance_ratio_))
plt.xlabel('номер компонент')
plt.ylabel('размер объясненной дисперсии')
plt.axvline(linewidth=4, color='r', linestyle = '--', x=10, ymin=0, ymax=1)
display(plt.show())
evr = pca_test.explained_variance_ratio_
cvr = np.cumsum(pca_test.explained_variance_ratio_)
pca_df = pd.DataFrame()
pca_df['Cumulative Variance Ratio'] = cvr
pca_df['Explained Variance Ratio'] = evr
display(pca_df.head(10))
```

GridSearch работает минут 30



```python
from sklearn.model_selection import RandomizedSearchCV 
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 30, num = 10)]
max_features = ['log2', 'sqrt']
max_depth = [int(x) for x in np.linspace(start = 5, stop = 20,num=5)]
min_samples_split = [int(x) for x in np.linspace(start = 5, stop = 15, num = 5)]
min_samples_leaf = [int(x) for x in np.linspace(start = 5, stop = 15, num = 5)]
bootstrap = [True, False]
param_dist = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
model = RandomForestClassifier()
rs = RandomizedSearchCV(model,param_dist,cv=3,n_jobs=-1,n_iter=10,random_state=42) 
#grid(RandomForestClassifier(),
                        #param_dist,
                        #cv = 3, 
                       # n_jobs=-1,
                    #random_state=42)
rs.fit(X_train_augm, y_train_augm)
rs.best_params_
```
