```python
list_genre=[]
list_genre= anime['genre'].values
list_genre[0:10]
```

```python
corpus=[]
for i in range(0,12294):
    first=list_genre[i]
    first = first.lower()
    first = first.replace(' ','')
    first = first.split(',')
    corpus.append(first)
```

```python
print(corpus)
```

```python
len(corpus)
```

```python
unique_corpus = []
for x in corpus:
    if x not in unique_corpus:
        unique_corpus.append(x)
```
