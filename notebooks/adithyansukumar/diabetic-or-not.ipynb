Now lets make a machine learning model,we will be using
1)logistic regression
2)svm
3)xgboost

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x=df.drop('Outcome',axis=1)
y=df['Outcome']
sc=StandardScaler()
x=sc.fit_transform(x)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
print("shape of x :{}".format(x.shape))
print("shape of y :{}".format(y.shape))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.metrics import classification_report,f1_score
from sklearn.metrics import plot_precision_recall_curve,average_precision_score
```

```python
def logistic_reg(x_train,x_test,y_train,y_test):
    lg=LogisticRegression(C=1,penalty='l2')
    lg.fit(x_train,y_train)
    lg_predict=lg.predict(x_test)
    lg_score=lg.decision_function(x_test)
    lg_avg_precision=average_precision_score(y_test,lg_score)
    print("classification report: \n",classification_report(y_test,lg_predict))    
    print("F1 score:{}".format(f1_score(y_test,lg_predict)))
    print("Average precision recall score :{}".format(lg_avg_precision))
    disp = plot_precision_recall_curve(lg, x_test, y_test)
    disp.ax_
```

```python
logistic_reg(x_train,x_test,y_train,y_test)
```

```python
def support_veector_machine(x_train,x_test,y_train,y_test):
    svc=SVC(kernel='rbf')
    svc.fit(x_train,y_train)
    svc_predict=svc.predict(x_test)
    svc_score=svc.decision_function(x_test)
    svc_avg_precision=average_precision_score(y_test,svc_score)
    print("classification report: \n",classification_report(y_test,svc_predict))    
    print("F1 score:{}".format(f1_score(y_test,svc_predict)))
    print("Average precision recall score :{}".format(svc_avg_precision))
    disp=plot_precision_recall_curve(svc,x_test,y_test)
    disp.ax_
    
```

```python
support_veector_machine(x_train,x_test,y_train,y_test)
```

```python
def xg_boost(x_train,x_test,y_train,y_test):
    xgb=XGBClassifier()
    xgb.fit(x_train,y_train)
    xgb_predict=xgb.predict(x_test)
    print("classification report: \n",classification_report(y_test,xgb_predict))    
    print("F1 score:{}".format(f1_score(y_test,xgb_predict)))
    disp=plot_precision_recall_curve(xgb,x_test,y_test)
    disp.ax_
```

```python
xg_boost(x_train,x_test,y_train,y_test)
```
