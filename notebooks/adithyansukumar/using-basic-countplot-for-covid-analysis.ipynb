```python
#importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#creating the dataframe
df=pd.read_csv("../input/covid19-patient-precondition-dataset/covid.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#as we can see the columns with dates are of type object lest convert them to date time format 
#many columns have null vlaues which are represented by 97,98,99,in the column date_died we can see 9999 99 99 as the date ,from this we can understand that 
#the patient is alive and not dead ,so lets replace the value 9999 99 99 with somthing like not dead or alive etc
#for  plotting purposes lets convert the values 1 and 2 to the respective strings
# but for some other ploting we may have to use numbers 
def clean_data(df):
    date_cols=['entry_date','date_symptoms']#we are not converting date_died column to date time type
    for col in date_cols:
        df[col]=pd.to_datetime(df[col],infer_datetime_format=True)
    df['date_died'].replace("9999-99-99","alive",inplace=True)
    
    for col in df.columns[:4]:
        df[col].replace([97,98,99],np.nan,inplace=True)
    for col in df.columns[5:]:
        df[col].replace([97,98,99],np.nan,inplace=True)
    df['sex'].replace([1,2],['female','male'],inplace=True)
    df['patient_type'].replace([1,2],['outpatient','inpatient'],inplace=True)
    for col in df.columns[6:8]:
        df[col].replace([1,2],['yes','no'],inplace=True)
    for col in df.columns[9:21]:
        df[col].replace([1,2],['yes','no'],inplace=True)
    df['covid_res'].replace([1,2,3],['positive','negative','awaiting results'],inplace=True)
    df['icu'].replace([1,2],['yes','no'],inplace=True)
    return df
    
```

```python
df=clean_data(df)
```
