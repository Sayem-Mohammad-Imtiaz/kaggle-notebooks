```python
df = pd.read_csv('../input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')
df
```

```python
df.info()
```

```python
df.gender.value_counts()
```

```python
df[df.gender=='Other']
```

```python
# drop row which gender = Other
df = df[df.gender != 'Other']
df
```

```python
# replacing NaN value in bmi with median of its column
df['bmi'] = df.bmi.fillna(df.bmi.median())
df
```

```python
#drop the outliers in bmi column
df = df[(df.bmi <= df.bmi.median() + 3 * df.bmi.std()) & (df.bmi >= df.bmi.median() - 3 * df.bmi.std())]
df
```

```python
# drop 'id' column 
df = df.drop('id', axis=1)
df
```

```python
# one-hot encoding 
columns_selected = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
df = pd.get_dummies(data=df, columns=columns_selected)
df
```

```python
# move stroke(target variable) to last column
new_cols = [col for col in df.columns if col != 'stroke'] + ['stroke']
df = df[new_cols]
df
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
```

```python
features = df.columns[:-1]
target = df.columns[-1]
X = df[features]
y = df[target]
```

```python
# train_test_split configuration
test_size = 0.2
random_state = 1234

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
```

```python
#create rf object
randomforest = RandomForestClassifier()
```

```python
cross_val_score(randomforest, X_train, y_train, cv=5).mean()
```

```python
randomforest.fit(X_train, y_train)
```

```python
y_pred = randomforest.predict(X_test)
```

```python
from sklearn.metrics import f1_score
```

```python
f1_score(y_test, y_pred, average='weighted')
```
