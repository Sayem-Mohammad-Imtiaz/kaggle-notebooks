```python
import numpy as np
import pandas as pd
import os
import seaborn as sn
import matplotlib.pyplot as plt
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data=pd.read_csv('/kaggle/input/insurance/insurance.csv')
```

```python
#encode categorical features - starting with sex - M/F 1/0
data['sex'].replace('female', 0, inplace=True)
data['sex'].replace('male', 1, inplace=True)
data.head()
```

```python
#encode smoker feature - Y/N 1/0
data['smoker'].replace('yes', 1, inplace=True)
data['smoker'].replace('no', 0, inplace=True)
data.head()
```

```python
#encode region feature - onehot - creates new dataframe encoding region. then concatenate the 2 dataframes 
region_dummy = pd.get_dummies(data['region'])
#data=pd.concat([data, region_dummy], axis=1)
data.drop('region', axis=1, inplace=True)
data.head()
```

```python
#check if sexes are balanced - pretty close! 
data['sex'].value_counts()
```

```python
data.hist('age')
data.hist('bmi')
data.hist('children')
data.hist('charges')
```

```python
#correlation matrix shows that age, smoking and bmi are positively correlated with charges 
#positive correlation between bmi and region=southeast? 
#negative correlation between bmi and region=northeast & northwest?
corr_matrix = data.corr()
plt.figure(figsize=(10,8))
sn.heatmap(corr_matrix, annot=True)
plt.show()
```

now time to test out some models to predict charges! 
starting with linear regression...

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.ensemble import RandomForestRegressor
```

```python
#linear regression
x = data.drop(['charges'], axis = 1)
y = data.charges

x_train,x_test,y_train,y_test = train_test_split(x,y, random_state = 0)
lr = LinearRegression().fit(x_train,y_train)

y_train_pred = lr.predict(x_train)
y_test_pred = lr.predict(x_test)

print(lr.score(x_test,y_test))
#pretty good benchmark result!
```

```python
#polynomial model
X = data.drop(['charges','northeast', 'northwest', 'southeast', 'southwest'], axis = 1)
Y = data.charges

quad = PolynomialFeatures (degree = 2)
x_quad = quad.fit_transform(X)

X_train,X_test,Y_train,Y_test = train_test_split(x_quad,Y, random_state = 0)

plr = LinearRegression().fit(X_train,Y_train)

Y_train_pred = plr.predict(X_train)
Y_test_pred = plr.predict(X_test)

print(plr.score(X_test,Y_test))
```

Not bad! seems like a pretty accurate predictor of medical costs.
