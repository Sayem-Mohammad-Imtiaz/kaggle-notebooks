```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler,MinMaxScaler
```

```python
data=pd.read_csv('../input/salary-data-simple-linear-regression/Salary_Data.csv')
data.head()
```

```python
data.plot(kind='box',subplots=True)
```

```python
sns.distplot(data['YearsExperience'])
```

```python
sns.distplot(data['Salary'])
```

```python
sns.regplot(data=data,x='YearsExperience',y='Salary')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y = train_test_split(data[['YearsExperience']],data['Salary'],test_size=0.35,random_state=12)
```

```python
lr=LinearRegression()
lr.fit(train_x,train_y)
predict=lr.predict(test_x)
print('MAE',mean_absolute_error(predict,test_y))
print('MSE',mean_squared_error(predict,test_y))
print('RMSE',np.sqrt(mean_squared_error(predict,test_y)))
print('R2',r2_score(predict,test_y))
sns.regplot(x=test_x,y=predict)
```
