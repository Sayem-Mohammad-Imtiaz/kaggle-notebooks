```python
dataset.dropna(axis=0, subset = ["RainTomorrow","RainToday"], inplace = True)
```

Creat X and Y

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

Taking care of missing data (numeric)

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 1:6])
X[:, 1:6] = imputer.transform(X[:, 1:6])

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 7:8])
X[:, 7:8] = imputer.transform(X[:, 7:8])

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:, 10:-1])
X[:, 10:-1] = imputer.transform(X[:, 10:-1])
```

taking care of missing data (categorical)

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='constant')
imputer.fit(X[:, 6:7])
X[:, 6:7] = imputer.transform(X[:, 6:7])

from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='constant')
imputer.fit(X[:, 8:10])
X[:, 8:10] = imputer.transform(X[:, 8:10])
```

Check our dataset

```python
x_check =pd.DataFrame(X)
x_check.isnull().sum()
```

# Encoding categorical data

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
print(X)

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [54,56,57])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
print(X)
# Encoding the independent Variable(Numeric)
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:,-1] = le.fit_transform(X[:,-1])
print(X)
```

# Encoding the Dependent Variable

```python

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

# Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
randomforset = RandomForestClassifier(n_estimators = 10, random_state = 0)
randomforset.fit(X_train, y_train)
```

```python
prediction = randomforset.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_test,prediction))
accuracy_score(y_test, prediction)
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
```

```python
prediction = xgb.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_test,prediction))
accuracy_score(y_test, prediction)
```
