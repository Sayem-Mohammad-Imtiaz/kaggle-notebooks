```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python

df=pd.read_csv('../input/customer-analytics/Train.csv')
df.head()
```

```python
df.Mode_of_Shipment.unique()
```

```python
Ship_df = df[df['Mode_of_Shipment'] == 'Ship']
len(Ship_df)/len(df)*100
```

****About 67% of the product shipment is through Ships, and next 16.15% is through flights and roads

```python
df['Reached.on.Time_Y.N'].value_counts()
```

```python
df['Reached.on.Time_Y.N'].value_counts().plot(kind='bar', color='green')
plt.title('Receival on time')
```

```python
df.Product_importance.value_counts()
```

```python
df.Product_importance.value_counts().plot(kind='bar', color='blue')
plt.ylabel("Total orders")
plt.title('order info as per product importance')
```

```python
ratings=df['Customer_rating'].value_counts()

sns.barplot(x=ratings.index, y=rating.values, alpha=0.8)
plt.title('Customer Ratings', fontsize=18, pad=10)
plt.ylabel('No of customers', fontsize=14)
plt.xlabel('Ratings', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=10)
plt.figure(figsize=(10, 7))
plt.show()
```

The company has product ratings of '1' and '2' in equal numbers as '3', '4' and '5' and hence need to downsize the products in these two ratings category.

#### Customer Orders by Gender

```python
df['Gender'].value_counts()
```

We see that gender are almost evenly distributed.


```python
gender = df['Gender'].value_counts()

plt.figure(figsize=(7, 5))

plot = gender.plot.pie(fontsize=16, autopct='%1.0f%%', labeldistance=1.2, radius=1.5)
plt.show()
```

There is a slight female dominance in the customer order history.

```python
print (df[["Gender", "Customer_care_calls"]].groupby(['Gender']).mean().sort_values(by='Customer_care_calls', ascending = False))
```

```python
print (df[["Gender", "Prior_purchases"]].groupby(['Gender']).mean())
```

Gender seems to have same relation with other features like 'Prior Purchases' and 'Customer ratings'

```python
# Gender by Customer care call, Customer ratings, Prior purchases and Reached on time.

sns.set_style('darkgrid')

f,axes= plt.subplots(2,2, figsize=(10,10))

sns.countplot(x='Customer_care_calls',hue='Gender',data=df,palette='RdBu_r', ax=axes[0,0])
sns.countplot(x='Customer_rating',hue='Gender',data=df,palette='RdBu_r', ax=axes[0,1])
sns.countplot(x='Prior_purchases',hue='Gender',data=df,palette='RdBu_r', ax=axes[1,0])
sns.countplot(x='Reached.on.Time_Y.N',hue='Gender',data=df,palette='RdBu_r', ax=axes[1,1])

sns.set_style('whitegrid')
```

#### Explore by Warehouse Block ( A, B,C,D and F)

```python
df['Warehouse_block'].value_counts()
```

We see that Warehouse block F holds highest number of products and Warehouses 'A' and 'B' host least number of products.

```python
warehouse_info = df['Warehouse_block'].value_counts()

plt.figure(figsize=(7, 5))

plot = warehouse_info.plot.pie(fontsize=16, autopct='%1.0f%%', labeldistance=1.2, radius=1.5)
plt.show()
```

#### Lets see how busy is the customer care call centres

```python
df['Customer_care_calls'].value_counts()
```

 We see that there is a high frequency of atleast 4 calls about shipping enquiry. The company needs to review if this is the expected call rate or not.

```python
call_rate=df['Customer_care_calls'].value_counts()
sns.barplot(x=call_rate.index, y=call_rate.values)
sns.color_palette("tab10")
plt.title('Number of calls by customers', fontsize=18, pad=10)
plt.ylabel('Number of customers', fontsize=14)
plt.xlabel('Number of calls', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=10)
plt.figure(figsize=(10, 7))
plt.show()
```

#### Explore by prior purchases

```python
df['Prior_purchases'].value_counts()
```

While there are customers above 3000 that had 3 purchases earlier, customers that had purchased more than 6 times are just within few hundreds. So the company needs to work upon increasing the prior purchase of customers

```python
prior_purchases=df['Prior_purchases'].value_counts()

sns.barplot(x=prior_purchases.index, y=prior_purchases.values)
sns.color_palette("tab10")
plt.title('Prior purchases by customers', fontsize=18, pad=10)
plt.ylabel('Number of customers', fontsize=14)
plt.xlabel('Number of prior purchases', fontsize=14)
plt.xticks(fontsize=14)
plt.yticks(fontsize=10)
plt.figure(figsize=(10, 7))
plt.show()
```
