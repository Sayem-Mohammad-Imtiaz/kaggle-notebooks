**Out of 215 students, 67 arent placed. So I'm creating a subset of the current dataframe that contains only the data of placed students for analysis purposes.**

```python
new_df = df[df['status'] == 'Placed']
new_df
```

```python
Placed_df = new_df.copy()
```

```python
### Here we get the details of the person who bagged the highest salary
```

```python
i = Placed_df.max()
print("The person who got the max salary is ", i[1])
print("The max salary is ", i[14])
print("Does the person have prior workex?", i[9])
print("What is the specialisation of the person with max salary?", i[11])
```

**Data Visualizations**

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
Placed_df.gender.value_counts().plot(kind='bar', color = 'green')
plt.ylabel("No of placements")
plt.title("Placements according to gender")
plt.show()
```

```python
Placed_df.specialisation.value_counts().plot(kind='bar')
plt.ylabel("No of students")
plt.title("Placements according to specialisation")
plt.show()
```

```python
Placed_df['degree_t'].value_counts().plot(kind='bar')
plt.ylabel("Degree background among placed students")
```

```python
Placed_df.groupby('gender')['salary'].mean()
Placed_df.groupby('gender')['salary'].mean().plot(kind='bar', color = 'red')
plt.ylabel("Salary")
plt.title("Average Salary of Male and Female")
plt.show()
```

```python
sns.countplot(x="workex", data=Placed_df, color="c")
```
