# There's No Null Values but There's values equal zero which can't happen


```python
#Replace value with 0 skin with the minimum skin length 
df['SkinThickness']=df['SkinThickness'].replace(0,10)

# DROP Columns that have Zeros 

df=df.drop(columns=['Insulin'])

```

```python
for i in df.columns:
    print(i)
    print(df[df[i] == 0].shape[0])
```

# Model 

```python
train, test = train_test_split(df, test_size=0.2, random_state=50)
x_train = train.drop(columns='Outcome')
y_train = train['Outcome']

x_val = test.drop(columns='Outcome')
y_val = test['Outcome']
```

# Logistic Model

```python
LogModel=LogisticRegression(penalty='l2',solver='liblinear')
LogModel.fit(x_train,y_train)
Train_Predict = LogModel.predict(x_train)
Val_Predict = LogModel.predict(x_val)


print('accuracy_score For Training : {}'.format(accuracy_score(y_train,Train_Predict)))
print('accuracy_score For Testing  : {}'.format(accuracy_score(y_val,Val_Predict)))
```

# KNN Model


```python
KNNModel = KNeighborsClassifier(n_neighbors = 5).fit(x_train, y_train)
KNN_Train_Predict=KNNModel.predict(x_train)
KNN_Val_Predict = KNNModel.predict(x_val)
print('accuracy_score For Training : {}'.format(accuracy_score(y_train,KNN_Train_Predict)))
print('accuracy_score For Testing  : {}'.format(accuracy_score(y_val,KNN_Val_Predict)))
```
