```python
hb = pd.read_csv('/kaggle/input/hotel-booking-demand/hotel_bookings.csv')
hb.head()
```

```python
hb.isnull().sum()
```

```python
# Cheking the persentage Null values in column
hb.agent.isnull().sum() / hb.agent.shape[0]
```

```python
hb.company.isnull().sum() / hb.company.shape[0]
# As company columns have NULL values larger tha 70% then we will drop that column
```

```python
hb.drop(columns = ['company','reservation_status_date'], inplace = True)
```

```python
hb.country.value_counts()
# as max no. od rows have PRT as country so we will fill that on NA place
```

```python
hb.country.fillna('PRT' , inplace = True)
```

```python
hb.children.value_counts()
# max chk in family dont have children
```

```python
hb.children.fillna(0.0 , inplace= True)
```

```python
hb.agent.value_counts()
# Agent have different values lets fill mean for NA
```

```python
agent_mean = int(hb['agent'].mean())
```

```python
hb.agent.fillna(agent_mean, inplace=True)
```

```python
# Chk for any other null values
hb.isnull().sum()
```

```python
# Now Data is cleaned 
# Lets apply LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
hb[hb.select_dtypes(include = ['object']).columns] = hb[hb.select_dtypes(include=['object']).columns].apply(le.fit_transform)
```

```python
# Now the data set is converted from Label to numeric data
hb.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (18,18))
ax = sns.heatmap(hb.corr())
```

```python
# Now define dependent and independent variables
hb_x = hb.drop('is_canceled' , axis =1)
hb_y = hb['is_canceled']
```

```python
# Split Train and Test data set
from sklearn.model_selection import train_test_split
```

```python
hb_x_train, hb_x_test, hb_y_train, hb_y_test = train_test_split(hb_x,hb_y, random_state=11)
```

```python
# Now import some more liberaries and fit our model
from sklearn.metrics import accuracy_score , confusion_matrix
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
rf.fit(hb_x_train,hb_y_train)
```

```python
pred_cncl = rf.predict(hb_x_test)
```

```python
pred_cncl
```

```python
accuracy_score(pred_cncl, hb_y_test)
```

```python
confusion_matrix(pred_cncl, hb_y_test)
```
