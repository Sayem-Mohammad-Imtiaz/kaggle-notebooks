```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import confusion_matrix,precision_score, recall_score
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression, Perceptron, SGDClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('../input/mushroom-classification/mushrooms.csv')
```

```python
train_df = df['class']
pred_df = df.drop('class',axis=1)
```

```python
pred_df = pd.get_dummies(pred_df, prefix='', prefix_sep='')
```

```python
y = train_df.values
x = pred_df.values
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
sc=StandardScaler()
X_train = sc.fit_transform(x_train)
X_test = sc.transform(x_test)
```

```python
model_nm, model_tt, model_rs, model_ps = [],[],[],[]
model_tp, model_fp, model_fn, model_tn = [],[],[],[]
```

```python
def con_mat():
    cm = confusion_matrix(y_test, y_pred)
    pssc = precision_score(y_test,y_pred, average= 'weighted')
    rssc = recall_score(y_test,y_pred, average= 'weighted')
    model_nm.append(model_name)
    model_tt.append(model_title)
    model_ps.append(pssc)
    model_rs.append(rssc)
    model_tp.append(cm[0][0])
    model_fp.append(cm[0][1])
    model_fn.append(cm[1][0])
    model_tn.append(cm[1][1])
```

```python
model_name = 'logr'
model_title = 'LR LogisticRegression'
logr = LogisticRegression(random_state=19)
logr.fit(X_train,y_train)
y_pred = logr.predict(X_test)
con_mat()
```

```python
model_name = 'sgd'
model_title = 'SGD StochasticGradientDescent'
sgd = SGDClassifier()
sgd.fit(X_train,y_train)
y_pred = sgd.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
con_mat()
```

```python
model_name = 'prc'
model_title = 'PRC Perceptron'
prc = Perceptron()
prc.fit(X_train,y_train)
y_pred = prc.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
con_mat()
```

```python
model_name = 'dtc'
model_title = 'DTC DecisionTree'
dtc = DecisionTreeClassifier(criterion = 'entropy')
dtc.fit(X_train,y_train)
y_pred = dtc.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
con_mat()
```

```python
model_name = 'gnb'
model_title = 'GNB Gaussian'
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
con_mat()
```

```python
model_name = 'svc'
model_title = 'SVC SupportVectorMachines'
svc = SVC(kernel='rbf')
svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
con_mat()
```

```python
model_name = 'knn'
model_title = 'KNN KNearistNeighbors'
knn = KNeighborsClassifier(n_neighbors=55, metric='minkowski')
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
con_mat()
```

```python
model_name = 'rfc'
model_title = 'RFC RandomForest'
rfc = RandomForestClassifier(n_estimators=30, criterion = 'entropy')
rfc.fit(X_train,y_train)
y_pred = rfc.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
con_mat()
```

```python
model_list = [model_nm, model_tt, model_rs, model_ps, model_tp, model_fp, model_fn, model_tn]
models_df = pd.DataFrame(model_list).transpose()
models_df.columns =  ['Model Name' , 'Model Title', 'Prsc' , 'Rcsc' ,'Tp', 'Fp', 'Fn', 'Tn']
```
