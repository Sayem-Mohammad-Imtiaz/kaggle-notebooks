```python
# Winner by Seasons

import datetime
data_by_date = data.sort_values(by="date")
data_by_date.head()

winners = []
i = 0
prev = 2008
years = pd.DatetimeIndex(data_by_date['date']).year

for year in years:
    if(year != prev):
        print(str(prev) + " : " + data_by_date.iloc[i-1,10])
        prev = year
    i += 1
print(str(prev) + " : " + data_by_date.iloc[-1,10])
```

```python
x=['Sunrisers Hyderabad', 'Mumbai Indians', 'Gujarat Lions',
    'Rising Pune Supergiant', 'Royal Challengers Bangalore',
    'Kolkata Knight Riders', 'Delhi Daredevils', 'Kings XI Punjab',
    'Chennai Super Kings', 'Rajasthan Royals', 'Deccan Chargers',
    'Kochi Tuskers Kerala', 'Pune Warriors', 'Rising Pune Supergiants', 'Delhi Capitals']

y = ['SRH','MI','GL','RPS','RCB','KKR','DC','KXIP','CSK','RR','SRH','KTK','PW','RPS','DC']

df = data.replace(x,y)
# df.head()
```

```python
df["winner"].value_counts()
```

# **Observation 1**

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(1,3,figsize = (20,8),sharey = True)
fig.suptitle('Winners', fontsize=16)


sns.countplot(ax = ax[0],
              x = "winner",
              order = df["winner"].value_counts().index,
              data = df)
ax[0].set_title("Match Winner")

sns.countplot(ax = ax[1],
              x = "toss_winner",
              order = df["toss_winner"].value_counts().index,
              data = df)
ax[0].set_title("Toss Winner")


toss_win = df[df["winner"] == df["toss_winner"]]
sns.countplot(ax = ax[2],
              x = "toss_winner",
              order = toss_win["winner"].value_counts().index,
              data = toss_win)
ax[0].set_title("Toss and Match Winner")
```

# **Observation 2**

```python
only_teams = df[["team1","team2","winner"]]

played_1 = only_teams["team1"].value_counts()
played_2 = only_teams["team2"].value_counts()
played = played_1 + played_2
winners = only_teams["winner"].value_counts()

match_vs_win = pd.DataFrame({
    "played" : played,
    "wins" : winners
}, index = winners.index)
match_vs_win['Win percentage']=((match_vs_win['wins']/match_vs_win['played'])*100).round()
match_vs_win
```
