It would appear men are at more risk of heart attacks. Lets see how well a predictor on age and sex would perform

```python
def classify_from_columns(X, y):
    knn_clf = KNeighborsClassifier()
    scores = cross_val_score(knn_clf, X, y, cv=5)
    #knn_clf.fit(X_train, Y_train)
    print(scores)
```

```python
X = pd.get_dummies(data_pd[["age", "sex"]], columns=["sex"])
classify_from_columns(X, data_pd["output"])
```

Lets add exercise induced angina

```python
X = pd.get_dummies(data_pd[["age", "sex","exng"]], columns=["sex","exng"])
classify_from_columns(X, data_pd["output"])
```

Perhaps chest pain could be another good indicator?

```python
X = pd.get_dummies(data_pd[["age", "sex", "exng", "cp"]], columns = ["sex", "exng", "cp"])
classify_from_columns(X, data_pd["output"])
```

```python
X = pd.get_dummies(data_pd[["age", "sex", "exng", "cp", "chol"]], columns = ["sex", "exng", "cp"])
classify_from_columns(X, data_pd["output"])
```

Interestingly, the addition of cholestoral levels as a feature appears to make the model behave worse. Lets follow our correlated features Series and add the next highest instead

```python
X = pd.get_dummies(data_pd[["age", "sex", "exng", "cp", "oldpeak"]], columns = ["sex", "exng", "cp"])
classify_from_columns(X, data_pd["output"])
```

```python
X = pd.get_dummies(data_pd[["age", "sex", "exng", "cp", "oldpeak", "thalachh"]], columns = ["sex", "exng", "cp"])
classify_from_columns(X, data_pd["output"])
```

The scores seem to be more clustered when thalachh suggesting this set of features is somewhat more consistent

```python
def forest_from_columns(X, y):
    clf = RandomForestClassifier()
    scores = cross_val_score(clf, X, y, cv=5)
    print(scores)
```

```python
X = pd.get_dummies(data_pd[["age", "sex", "exng", "cp", "oldpeak", "thalachh"]], columns = ["sex", "exng", "cp"])
forest_from_columns(X, data_pd["output"])
```

Immediate improvement on using Random Forest, not unexpected
