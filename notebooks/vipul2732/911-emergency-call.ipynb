```python
# from the title column grab the reason in another column and show the most common reason for the 911 calls
```

```python
#x = df['title'].loc[0]
```

```python
#x.split(':')[0]
```

```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
```

```python
df['Reason'].value_counts()  ## EMS is the most common reason
```

```python
sns.countplot(x='Reason', data=df) # seaborn to show the most common reasons
```

```python
type(df['timeStamp'].iloc[0])
```

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp']) ## convert timestamp data from string to daytime
```

```python
type(df['timeStamp'].iloc[0])
```

```python
time = df['timeStamp'].iloc[0]
time.hour
```

```python
# create three new coloumns for defening the time , month and day
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['day_of_week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
df.head()
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
```

```python
df['day_of_week'] = df['day_of_week'].map(dmap) # convert day of week form num to actual day
```

```python
df.head()
```

```python
sns.countplot(x='day_of_week', data =df, hue = 'Reason')  ## Day wise reason for the call
plt.legend(bbox_to_anchor=(1.05, 1))    ## reasons >  EMS>traffic>Fire   ## call frequency > Friday=Mon=Tue > Thu> Wed> sat> sun
```

```python
sns.countplot(x='Month', data =df, hue = 'Reason')  ## Monthwise wise reason for the call
plt.legend(bbox_to_anchor=(1.3, 1))    ## month first and seventh had highest calls
```

```python
## in count plot we missed some value from months 8 to 11, so try to fill thsi
```

```python
df["date"] = df['timeStamp'].apply(lambda t: t.date())  ## add date column
```

```python
df.head()
```

```python
df.groupby('date').count()['lat'].plot()
plt.tight_layout()
```

```python
df[df['Reason']=='Fire'].groupby('date').count()['lat'].plot()    #Fire
plt.tight_layout()
```

```python
df[df['Reason']=='Traffic'].groupby('date').count()['lat'].plot()   #Traffic
plt.tight_layout()
```

```python
df[df['Reason']=='EMS'].groupby('date').count()['lat'].plot()  ##EMS
plt.tight_layout()
```

```python
day_hour = df.groupby(by=['day_of_week','Hour']).count()['Reason'].unstack()
```

```python
sns.heatmap(day_hour, cmap ='coolwarm')   ## most of the emergency call happening in the mid to late eveneing  # very less in the early morning 
```

```python
sns.clustermap(day_hour, cmap ='coolwarm')
```
