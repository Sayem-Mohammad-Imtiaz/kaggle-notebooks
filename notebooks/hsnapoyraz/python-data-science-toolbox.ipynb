
SCOPE<br>
What we need to know about scope:<br>

global: defined main body in script<br>
local: defined in a function<br>
built in scope: names in predefined built in scope module such as print, len<br>

```python
x = 2
def f():
    x = 3
    return x
print(x)      # x = 2 global scope
print(f())    # x = 3 local scope
```

```python
x = 5              # global scope x  
def f():
    y = 2*x        # there is no local scope x
    return y
print(f())         # it uses global scope x  
# First local scope searched, then global scope searched, if two of them cannot be found lastly built in scope searched.
```


NESTED FUNCTION<br>
* function inside function.
* There is a LEGB rule that is search local scope, enclosing function, global and built in scopes, respectively.

```python
#nested function
def square():
    """ return square of value """
    def add():
        """ add two local variable """
        x = 2
        y = 3
        z = x + y
        return z
    return add()**2
print(square())  
```

DEFAULT and FLEXIBLE ARGUMENTS<br>

1. Default argument example:<br>

def f(a, b=1):<br>
""" b = 1 is default argument"""

2. Flexible argument example:<br>

def f(*args):<br>
 """ *args can be one or more"""<br>
 
 def f(** kwargs)<br>
 """ **kwargs is a dictionary"""

```python
# default arguments
def f(a, b = 1, c = 2):
    y = a + b + c
    return y
print(f(5))
# what if we want to change default arguments
print(f(5,4,3))
```

```python
# flexible arguments *args
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,3,5,7)

# flexible arguments **kwargs that is dictionary
def f(**kwargs):
    """ print key and value of dictionary"""
    for key, value in kwargs.items():              
        print(key, " ", value)
f(country = 'spain', capital = 'madrid', population = 123456)
```

LAMBDA FUNCTION<br>
Faster way of writing function

```python
#user defined function
def square(x):
    return x**2
print(square(5))

# lambda function
square = lambda x: x**2     # where x is name of argument
print(square(4))

tot = lambda x,y,z: x+y+z   # where x,y,z are names of arguments
print(tot(1,2,3))
```

ANONYMOUS FUNCTİON<br>

Like lambda function but it can take more than one arguments.<br>

* map(func,seq) : applies a function to all the items in a list

```python
#anonymous function
number_list = [1,2,3]
y = map(lambda x:x**2,number_list)
print(list(y))
```

```python
#anonymous function
mylist = [1,2,3,4,5]
x = map(lambda x: x*5, mylist)
print(list(x))
```

ITERATORS<br>

* iterable is an object that can return an iterator<br>
* iterable: an object with an associated iter() method<br>
example: list, strings and dictionaries<br>
* iterator: produces next value with next() method

```python
# iteration example
name = "ronaldo"   #string
it = iter(name)    # use iter() method
print(next(it))    # print next iteration
print(next(it))
print(*it)         # print remaining iteration
```

zip(): zip lists

```python
# zip example
list1 = [1,2,3,4]     # list
list2 = [5,6,7,8]
z = zip(list1,list2)  # use zip() method
print(type(z))
print(z)
print("------")

z_list = list(z)
print(type(z))
print(z_list)
print("------")

un_zip = zip(*z_list)            # used to "*" unzip
un_list1,un_list2 = list(un_zip) # unzip returns tuple
print(type(un_list2)) 
print(un_list1)
print(un_list2)      
```

LIST COMPREHENSİON<br>
One of the most important topic of this kernel<br>

* list comprehension: collapse for loops for building lists into a single line<br>
Ex: num1 = [1,2,3] and we want to make it num2 = [2,3,4]. This can be done with for loop. However it is unnecessarily long. We can make it one line code that is list comprehension.

```python
# Example of list comprehension
num1 = [1,2,3]
num2 = [i + 1 for i in num1 ]
print(num2)
```

[i + 1 for i in num1 ]: list of comprehension<br>
i +1: list comprehension syntax<br>
for i in num1: for loop syntax<br>
i: iterator<br>
num1: iterable object<br>

```python
# Conditionals on iterable
num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)
```
