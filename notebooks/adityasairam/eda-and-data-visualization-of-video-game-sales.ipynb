```python
features_with_na=[features for features in df.columns if df[features].isnull().sum()>1]

for feature in features_with_na:
    print(feature, np.round(df[feature].isnull().mean(), 4),  ' % missing values')##Finding out the missing values
```

<b>As we can see, the percentage of Nan values present in the dataset is very less.Hence we can just drop them.

```python
df.dropna(inplace=True)
```

<b><h4>Splitting the datafame with respect to time periods.I am considering three time periods for simplicity</h4>

```python
old_times_df = df[(df.Year >= 1980) & (df.Year < 2003)] 
mid_times_df = df[(df.Year >=2003) & (df.Year < 2015)]
recent_times_df = df[df.Year >=2015]
```

```python
old_times_df['Platform'].describe()
```

```python
mid_times_df['Platform'].describe()
```

```python
recent_times_df['Platform'].describe()
```

```python
old_times_df.groupby('Platform')['Global_Sales'].sum().plot(kind = "bar")
```

```python
mid_times_df.groupby('Platform')['Global_Sales'].sum().plot(kind = "bar")
```

```python
recent_times_df.groupby('Platform')['Global_Sales'].sum().plot(kind = "bar")
```

<h4> From the above visualizations, We can see that different platforms have dominated the market during different periods of time.In short,
 
<b> 1. From 1980-2003 : PS,PS2,SNES<br>
<b>  2. From 2003-2015 : PS2,PS3,Wii,X360,DS<br>
<b>  3. From 2015-Present : Ps4,Xone<br>
