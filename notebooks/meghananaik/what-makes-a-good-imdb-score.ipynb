**Creating a mini dataframe for all rows with IMDB score above 7.6**

```python
score_7 = df[df['IMDB Score']>=7.6]
score_7.head()
```

**Breaking up the Genre and Language columns where some rows have multiple values **

```python
## Breaking up the genres and Languages
temp_df = score_7
s = temp_df['Genre'].str.split('/').apply(pd.Series, 1).stack()
t = temp_df['Language'].str.split('/').apply(pd.Series, 1).stack()
s.index = s.index.droplevel(-1) # to line up with df's index
t.index = t.index.droplevel(-1)
s.name = 'Genre_split'
t.name = 'Language_split'
temp_df = temp_df.join(s)
temp_df = temp_df.join(t)
score_7 = temp_df
score_7.head()
```

**Which Languages have a score of 7.6 or above?**

```python
score_7['Language_split'].value_counts()
```

```python
plt.figure(figsize = (20,12))
sns.countplot(x="Language_split", data= score_7)
```

**Which Genres have a score of 8 or above?**

```python
plt.figure(figsize = (35,12))
#sns.barplot(x = 'IMDB Score', y = 'Language',hue ='Genre', data = score_8 )
sns.countplot(x="Genre_split", data= score_7)
```

Documentaries have good IMDB scores compared to other Genres.

```python
score_7['year'] = pd.DatetimeIndex(score_7['Premiere']).year
```

****Which year had most hits ? ****

```python
plt.figure(figsize = (35,12))
#sns.barplot(x = 'IMDB Score', y = 'year',hue ='Genre', data = score_8 )
sns.countplot(x="year", data= score_7)
```
