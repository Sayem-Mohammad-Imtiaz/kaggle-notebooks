```python
df2=df1.copy()
```

```python
df2.head()
```

```python
def isfloat(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
x=df2[~df2["total_sqft"].apply(isfloat)]
```

```python
x
```

```python
def convert_sqft_float(x):
    token=x.split('-')
  
    if len(token)==2:
        return ((float(token[0])+float(token[1]))/2)
    try:
        return float(x)
    except:
        return None
```

```python
df2["total_sqft"]=df2["total_sqft"].apply(convert_sqft_float)
```

```python
df2.head()
```

```python
df3=df2.copy()
```

```python
df3['total_sqft']=df3['total_sqft'].fillna(df3['total_sqft'].median())
```

```python
df3.info()
```

```python
df3["price_per_sqft"]=df3["price"]*100000/df3["total_sqft"]
```
