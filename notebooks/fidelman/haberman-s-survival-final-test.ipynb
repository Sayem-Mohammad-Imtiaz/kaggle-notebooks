```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency

%matplotlib inline
```

```python
df = pd.read_csv("../input/habermans-survival-data-set/haberman.csv", header=None)
df = df.drop([1,2], axis=1)
df.columns = ["age", "survival_status"]
df = df[df.age <= 65]
df
```

```python
contigency = pd.crosstab(df['survival_status'], df['age'])
contigency
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(contigency, annot=True)
```

```python
c, p, dof, expected = chi2_contingency(contigency) 
c_critical = 50
c, c_critical
```

retain the null hypothesis

```python
cramers_v = math.sqrt(c / contigency.sum().sum())
cramers_v
```

The result is moderate. The fields are moderately associated.

```python
contigency_age_groups = pd.DataFrame()
age_groups = pd.Series([(0, 40),(40, 56), (56, 66)])
for min_ag, max_ag in age_groups:
    contigency_age_groups[f'({min_ag},{max_ag})'] = df[(min_ag <= df.age) & (max_ag > df.age)].groupby(["survival_status"]).count().age
contigency_age_groups
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(contigency_age_groups, annot=True)
```

```python
c, p, dof, expected = chi2_contingency(contigency_age_groups) 
c_critical = 5.991
c, c_critical
```

reject the null

```python
cramers_v = math.sqrt(c / contigency_age_groups.sum().sum())
cramers_v
```

The result is weak. Although the result is statistically significant, the fields are only weakly associated.
