```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
import numpy as np
 
style.use('fivethirtyeight')

data = pd.read_csv("../input/ipl-complete-dataset-20082020/IPL Ball-by-Ball 2008-2020.csv")
row_count = len(data.index)
col_count = len(data.columns)
#row_count_cols = data.count
print("The number of rows in file:" + str(row_count))
print("The number of columns in file:" + str(col_count))
data.head()

```

```python
# No of unique values in columns 

for col in data.columns:
    print(col + " : " + str(data[col].nunique()))
    
```

```python
# Total runs by team wise  
df_agg = data.groupby(['batting_team']).agg({'total_runs':[sum],'over':[sum]}).reset_index()

file_obj  = open("./runs.csv","w+")
df_agg.to_csv("./runs.csv",index=False)
df_agg_new_temp = pd.read_csv("./runs.csv",skiprows=[1])
df_agg_new = df_agg_new_temp.sort_values(by='total_runs',ascending=False)

# Total runs by team wise - Plot
objects = df_agg_new['batting_team']
x_pos = np.arange(len(objects))
performance = df_agg_new['total_runs']
performance

plt.bar(x_pos, performance, align='center', alpha=1)
plt.xticks(x_pos, objects, rotation=90)
plt.ylabel('Total Runs')
plt.title('IPL Total Runs between 2008 and 2020')
plt.show()

```

```python
# Top 10 bowlers who bowled max overs

#data.sort_values(by=['over'],ascending=False).head(10)
df_agg_bowler_tmp  = data.groupby(['bowler']).agg({'over':[sum]}).reset_index()
file_obj  = open("./df_agg_bowler_tmp.csv","w+")
df_agg_bowler_tmp.to_csv("./df_agg_bowler_tmp.csv",index=False)
df_agg_bowler = pd.read_csv("./df_agg_bowler_tmp.csv",skiprows=[1])
df_agg_bowler_top_10 = df_agg_bowler.sort_values(by='over',ascending=False).head(10)

# Total runs by team wise - Plot
objects = df_agg_bowler_top_10['bowler']
x_pos = np.arange(len(objects))
performance = df_agg_bowler_top_10['over']
#performance

plt.bar(x_pos, performance, align='center', alpha=1)
plt.xticks(x_pos, objects, rotation=90)
#plt.axhline(y=df_agg_bowler['over'].mean(), color='g', linestyle='-',label="horizontal")
plt.ylabel('Total Overs')
plt.title('IPL Top 10 bowler based on number of \n overs bowled between 2008 and 2020')


```

```python
data_top10_batsman_split = pd.DataFrame(data,columns=['batsman','batsman_runs'])
data_top10_batsman_bfr_sort_tmp = data_top10_batsman_split.groupby(['batsman']).agg({'batsman_runs':[sum]}).reset_index()
file_obj  = open("./df_agg_bowler_tmp.csv","w+")
data_top10_batsman_bfr_sort_tmp.to_csv("./data_top10_batsman.csv",index=False)
data_top10_batsman=pd.read_csv("./data_top10_batsman.csv",skiprows=[1]).sort_values(by='batsman_runs',ascending=False).head(10).set_index('batsman')
data_top10_batsman.plot.bar(alpha=1,rot=90,title='IPL Top 10 batsman based on number of \n runs scored between 2008 and 2020')

```

```python
df_bowling_team_temp = data.groupby(['batting_team','bowling_team']).agg({'total_runs':[sum]}).reset_index()
file_obj  = open("./df_bowling_team_temp.csv","w+")
df_bowling_team_temp.to_csv("./df_bowling_team_temp.csv",index=False)
df_bowling_team = pd.read_csv("./df_bowling_team_temp.csv",skiprows=[1])


for x in df_bowling_team['bowling_team'].unique():
    filter_chennai = df_bowling_team['bowling_team']==x
    display_chennai = df_bowling_team[filter_chennai].sort_values(by=['total_runs'],ascending=False)
    #print(display_chennai)
    #Total runs by team wise - Plot
    objects_chennai_temp = display_chennai['bowling_team'] + " against " + display_chennai['batting_team'] 
    objects_chennai = display_chennai['batting_team'] 
    x_pos = np.arange(len(objects_chennai))
    performance = display_chennai['total_runs']
    plt.bar(x_pos, performance, align='center', alpha=1)
    plt.xticks(x_pos, objects_chennai, rotation=90)
    plt.ylabel('Total Runs')
    plt.title(x + " given runs vs ")
    plt.show()
  



```
