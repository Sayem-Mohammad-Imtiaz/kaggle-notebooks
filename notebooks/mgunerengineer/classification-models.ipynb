**Normalization**

```python
y = df.target.values
x_data = df.drop(['target'],axis=1)
x = (x_data-np.min(x_data))/(np.max(x_data)-np.min(x_data))
```

**Train and Test Split**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

**Fit Logistic Regression Model**

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
accuracy = lr.score(x_test, y_test)
print('Accuracy of Logistic Regression Model: ',accuracy)
```

Accuarcy of Logistic Regression Model is 0.85.

**Fit SVM Classification Model**

```python
from sklearn.svm import SVC
svm = SVC(random_state=42)
svm.fit(x_train, y_train)
accuracy2 = svm.score(x_test, y_test)
print('Accuracy of SVM Model: ',accuracy2)
```

Accuracy of SVM Model is 0.836.

**Fit KNN Classification Model**

```python
accuracy3 = []
from sklearn.neighbors import KNeighborsClassifier
for each in range (1,15):
    knn =  KNeighborsClassifier(n_neighbors=each)
    knn.fit(x_train, y_train)
    accuracy3.append(knn.score(x_test, y_test))
#print('Accuracy of KNN Model: ',accuracy3)
```

```python
plt.plot(range(1,15),accuracy3)
plt.xlabel('Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

As seen in figure, Accuracy is high when neighbors was 3.

```python
np.max(accuracy3)
```

```python
np.argmax(accuracy3)
```

```python
print('Accuracy of KNN Model: ',accuracy3[2])
```

Accuracy of KNN Model is 0.836 when n_neighbors was 3.

**Fit Naive Bayes Classification Model**

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train, y_train)
accuracy4 = nb.score(x_test, y_test)
print('Accuracy of NB Classification Model: ',accuracy4)
```

Accuracy of Naive Bayes Classification Model is 0.868.

**Fit Decision Tree Classification Model**

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
accuracy5 = dt.score(x_test, y_test)
print('Accuracy of DT Classification Model: ',accuracy5)
```

Accuracy of Decision Tree Classification Model is 0.836.

**Fit Random Forest Classification Model**

```python
accuracy6 = []
from sklearn.ensemble import RandomForestClassifier
for each in range(1,150):
    rf = RandomForestClassifier(n_estimators=each, random_state=42)
    rf.fit(x_train, y_train)
    accuracy6.append(rf.score(x_test, y_test))

#print('Accuracy of RF Classification Model: ',accuracy6)
```

```python
plt.plot(range(1,150),accuracy6)
plt.xlabel('N_Estimators')
plt.ylabel('Accuracy')
plt.show()
```

As seen in figure, Accuracy is high when n_estimators was 49.

```python
np.max(accuracy6)
```

```python
np.argmax(accuracy6)
```

Accuracy of Random Forest Classification Model is 0.885 when n_estimators was 49.
