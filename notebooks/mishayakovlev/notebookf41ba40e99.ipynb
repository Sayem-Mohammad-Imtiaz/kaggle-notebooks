```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Data = pd.read_csv("/kaggle/input/tv-shows-on-netflix-prime-video-hulu-and-disney/tv_shows.csv")
Data = Data.drop(['Unnamed: 0'],axis=1)
```

```python
Data.head(10)
```

```python
Data.describe()
```

```python
Data.info()
```

```python
plt.figure(figsize=(16,12))
plt.title("Corr map",size=30)
sns.heatmap(Data.corr(),square=True,annot=True)
```

```python
byage = Data['Age'].value_counts()
byage['Unknow'] = Data[Data['Age'].isnull()].shape[0]
```

```python
plt.figure(figsize=(12,8))
plt.style.use(['seaborn-muted'])
plt.bar(list(byage.index),list(byage.values))
plt.title('By age',size=30)
plt.ylabel("Count of films",size=18)
```

```python
plt.figure(figsize=(12,8))
byyear = Data.groupby(by=['Year']).count().iloc[:,0]
plt.plot(byyear.index,byyear.values)
plt.xlabel("Year")
plt.ylabel("Count")
plt.title("Number of films by year")
```

```python
plt.figure(figsize=(12,8))
plt.boxplot(Data[Data['IMDb'].isnull() == False]['IMDb'])
plt.title("Boxplot of IMDb")
None
```

```python
plt.figure(figsize=(16,12))
plt.hist(Data[Data['IMDb'].isnull() == False]['IMDb'],bins=20)
None
```

```python
ls = Data[Data["Rotten Tomatoes"].isnull() == False]["Rotten Tomatoes"].apply(lambda x: int(x[:-1]))
```

```python
plt.figure(figsize=(16,12))
plt.boxplot(ls)
plt.ylim(0,11)
plt.yticks(np.arange(0, 105, 10))
plt.title("Rotten Tomatoes")
None
```

```python
plt.figure(figsize=(12,8))
plt.hist(ls,bins=max(ls)-min(ls),color='orange')
None
```
