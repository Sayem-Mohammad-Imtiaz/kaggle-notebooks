# Bar Plot Seaborn

Show point estimates and confidence intervals as rectangular bars.
 A bar plot represent an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainty around that estimate using error bars

```python
bar = sns.barplot(x='race', y='math score', data=df)
plt.xlabel('Race')
plt.ylabel('Math Score')
plt.title('Math Score with Race')
```

```python
bar = sns.barplot(x='race', y='math score', hue='gender', data=df, palette='autumn')
plt.title('Mathe Score with Race and Gender')
plt.xlabel('Race')
plt.ylabel('Math Score')
```

```python
bar = sns.barplot(x='race', y='math score', hue='parental level of education',
                 data = df, palette = 'rainbow')
plt.ylabel('Mathe Score')
plt.xlabel('Race')
plt.title('Mathe Score with Race and Parental Level of Education')
```

```python
bar = sns.barplot(x='race', y='reading score', hue='gender', data = df,
                 )
plt.ylabel('Reading Score')
plt.xlabel('Race')
plt.title("Reading Score with Race")
```

# Bar plot Ploty 

```python
df_rece_math=df.groupby(by =['race'])['reading score'].mean().to_frame().reset_index().rename(columns={'race':'Race','reading score':'Reading Score'})

fig = go.Figure(go.Bar(
    y=df_rece_math['Race'],x=df_rece_math['Reading Score'],orientation="h",
    marker={'color': df_rece_math['Reading Score'], 
    'colorscale': 'sunsetdark'},  
    text=df_rece_math['Reading Score'],
    textposition = "outside",
))
fig.update_layout(title_text='Mean Reading Score',xaxis_title="Race",yaxis_title="Reading Score",title_x=0.5)
fig.show()
```

```python
df_rece_math=df.groupby(by =['race'])['reading score'].mean().to_frame().reset_index().rename(columns={'race':'Race','reading score':'Reading Score'})

fig = go.Figure(go.Bar(
    x=df_rece_math['Race'],y=df_rece_math['Reading Score'],
    marker={'color': df_rece_math['Reading Score'], 
    'colorscale': 'Viridis'},  
    text=df_rece_math['Reading Score'],
    textposition = "outside",
))
fig.update_layout(title_text='Mean Reading Score',xaxis_title="Race",yaxis_title="Reading Score",title_x=0.5)
fig.show()
```

# Pie Plot Seaborn

```python
labels = df.race.value_counts().index
colors = ['red', 'green', 'blue','yellow', 'orange']
explode = [0.1,0,0,0,0]
sizes = df.race.value_counts().values

# visulize

plt.figure(figsize=(7,7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Race', color='black', fontsize=15)
```
