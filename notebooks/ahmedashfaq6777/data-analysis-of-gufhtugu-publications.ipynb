```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load data and Preprocessing

```python
data = pd.read_csv('../input/gufhtugu-publications-dataset-challenge/GP Orders - 5.csv')
data['Order Date & Time'] = pd.to_datetime(data['Order Date & Time'])
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

# Remove null Values

```python
data.dropna(inplace=True)
```

```python
data.isna().sum()
```

# Check Unique Values

```python
data['Order Status'].unique()
```

```python
data['Payment Method'] = data['Payment Method'].replace({'Cash on Delivery (COD)':'Cash on delivery'})
data['Payment Method'].unique()
```

# Check Categorical columns and Numericals coulmns

```python
cat_col = []
num_cols = []
for i in data.columns:
    if data[i].dtype == 'object':
        cat_col.append(i)
    else:
        num_cols.append(i)
```

```python
cat_col
```

```python
num_cols
```

```python
data['Order Date & Time'] = pd.to_datetime(data['Order Date & Time'])
```

```python
data['day'] = data['Order Date & Time'].dt.day_name()
data['month'] = data['Order Date & Time'].dt.month
data['month_name'] = data['Order Date & Time'].dt.month_name()
data['Year'] = data['Order Date & Time'].dt.year
data['Quarter'] = data['Order Date & Time'].dt.quarter
```

```python
data
```

# Exploratory Data Analysis

# Order Status Insights

```python
sns.set_style('whitegrid')
plt.title("Order Status Frequency")
sns.countplot(x='Order Status', data=data)
```

# Payment Method Frequency with Order Status

```python
plt.title("Payment Method Frequency with Order Status")
sns.barplot(x = 'Payment Method', y = 'Order Number', hue='Order Status', data=data)
```

# Compare day with order Status

```python
sns.barplot(x = 'day', y = 'Order Number',data=data, hue='Order Status')
```

```python
sns.countplot(x='Year', data=data)
```

```python
sns.countplot(x='day', data= data)
```

# Top 10 Cities by Order

```python
plt.title("Top 10 Cities by order")
plt.figure(figsize=(30, 10))
sns.countplot(x = 'City', data= data, order = data['City'].value_counts().iloc[:30].index)
```

# Top 10 Books

```python
plt.title("Top 10 Books")
data[data['Order Status']=='Completed']['Book Name'].value_counts().iloc[:10].sort_values().plot.barh()
```
