```python
ramen_drop_unrated = ramen.copy()
ramen_convert_unrated = ramen.copy()
```

```python
# Drop Unrated Ramen
ramen_drop_unrated = ramen_drop_unrated[ramen_drop_unrated["Stars"] != "Unrated"]
# ramen_drop_unrated.drop(ramen_drop_unrated[ramen_drop_unrated["Stars"] == "Unrated"].index, inplace=True)
```

```python
# Convert Unrated Ramen to 0
# ramen_convert_unrated.loc[ramen_drop_unrated["Stars"] == "Unrated", "Stars"] = 0
ramen_convert_unrated["rating"] = ramen_convert_unrated["Stars"].str.replace("Unrated", "0")
```

```python
ramen_drop_unrated["rating"] = ramen_drop_unrated["Stars"].astype(str).astype(float)
ramen_convert_unrated["rating"] = ramen_convert_unrated["rating"].astype(str).astype(float)
```

```python
ramen_drop_unrated["rating"].max()
```

```python
ramen_drop_unrated.groupby("Style")["rating"].mean()
```

```python
ramen_convert_unrated.groupby("Style")["rating"].mean()
```

```python
ramen_drop_unrated.groupby("Country")["rating"].mean().sort_values()
```

```python
ramen_drop_unrated.groupby("Brand")["rating"].count().sort_values(ascending=False)[:25]
```

```python
ramen_drop_unrated.groupby("Brand").agg({"rating": ["mean", "count"]}).sort_values([("rating", "mean")], ascending=False)[:25]
```
