## 1.1. O QUE É O BRICS

O BRICS é o agrupamento formado por cinco grandes países emergentes - Brasil, Rússia, Índia, China e África do Sul - que, juntos, representam cerca de 42% da população, 23% do PIB, 30% do território e 18% do comércio mundial.

 

O acrônimo BRIC foi cunhado em 2001 pelo banco de investimentos Goldman Sachs, para indicar as potências emergentes que formariam, com os Estados Unidos, as cinco maiores economias do mundo no século XXI. Em 2006, os países do BRIC deram início ao diálogo que, desde 2009, tem lugar nos encontros anuais de chefes de Estado e de Governo. Em 2011, com o ingresso da África do Sul, o BRICS alcançou sua composição definitiva, incorporando um país do continente africano.

 

Desde o início de seu diálogo, os países do agrupamento buscaram estabelecer governança internacional mais condigna com seus interesses nacionais, por meio, por exemplo, da reforma de cotas do Fundo Monetário Internacional, que passou a incluir, pela primeira vez, Brasil, Rússia, Índia e China entre os maiores cotistas.

 

Ao longo de sua primeira década, o BRICS desenvolveu cooperação setorial em diferentes áreas, como ciência e tecnologia, promoção comercial, energia, saúde, educação, inovação e combate a crimes transnacionais. Atualmente, essa cooperação setorial, que abrange mais de 30 áreas, traz importantes benefícios concretos para as populações dos países do agrupamento. É o caso da Rede de Pesquisa em Tuberculose, que visa a introduzir medicamentos e diagnósticos de qualidade a preços acessíveis.

 

Na última Cúpula sediada no Brasil, em 2014, foram criadas as instituições de Fortaleza: o Novo Banco de Desenvolvimento (NDB) e o Arranjo Contingente de Reservas (ACR). Até o momento, o NDB já aprovou mais de 8 bilhões de dólares em projetos de financiamento de infraestrutura e de energia renovável nos países do BRICS. O ACR, igualmente, encontra-se operacional e consiste em importante mecanismo de estabilidade financeira dos países em caso de desequilíbrio em seus balanços de pagamentos.

 

Além dos encontros presidenciais (cúpula e encontro informal à margem do G20), o BRICS organiza, por meio de sua presidência rotativa, cerca de 100 reuniões anuais, entre as quais cerca de 15 ministeriais e dezenas de encontros de altos funcionários, eventos técnicos, bem como reuniões nas áreas de cultura, educação e esporte.

 

Ao longo de 2019, o Brasil exercerá a presidência de turno do BRICS. A ênfase da presidência brasileira será na promoção de ciência, tecnologia e inovação; da economia digital; do aumento dos contatos entre o setor produtivo e o NDB; e no reforço da cooperação no combate a crimes transnacionais. Além disso, estão programados dezenas de eventos acadêmicos, esportivos, culturais e artísticos ao longo de todo o ano.

## 1.2. Principais áreas de cooperação
 
A primeira década de cooperação do BRICS foi marcada pela rápida ampliação dos temas tratados pelos parceiros.  Desde sua primeira cúpula, em 2009, o BRICS estabeleceu mais de 30 áreas de cooperação. Destacam-se, entre elas, a cooperação econômico-financeira; em saúde; ciência, tecnologia e inovação; segurança e empresarial. Em conjunto, essas iniciativas resultaram na conformação de importante patrimônio de realizações que visa a gerar benefícios concretos para nossas sociedades.¹

### 1.2.1. Econômico-Financeira

 O Novo Banco de Desenvolvimento do BRICS (NDB), cuja criação remonta à Cúpula de Fortaleza (2014), é uma das expressões mais concretas da cooperação econômico-financeira no âmbito do agrupamento. Por intermédio do banco, busca-se mobilizar recursos para incrementar o aporte de investimentos em infraestrutura e desenvolvimento sustentável nos países do BRICS, bem como em outras economias emergentes.

O capital subscrito inicial do banco é de US$50 bilhões, dos quais US$ 10 bilhões serão integralizados em partes iguais pelos cinco países até 2022. Tendo recebido classificação de risco AA+ da Fitch e da Standard & Poor's (S&P), o banco viabilizará aos países membros acesso a financiamento, com custo de captação reduzido, facilitando o cumprimento de sua finalidade institucional.

A sede do banco localiza-se em Xangai, na China, e o primeiro escritório regional, em Johanesburgo, na África do Sul, já está em operação. Em 2019, será inaugurado o Escritório Regional das Américas do NDB, com sede em São Paulo e representação em Brasília. O Escritório tem como objetivo prospectar e elaborar projetos a serem financiados pelo NDB no Brasil e na região.

Arranjo Contingente de Reservas (ACR)

O ACR constitui importante mecanismo de apoio macroeconômico para os países do BRICS. Esse arranjo visa a respaldar os países membros, especificamente em eventuais cenários de crise em seus balanços de pagamentos.

O total de recursos alocados inicialmente para o ACR totalizará US $ 100 bilhões, com os seguintes compromissos individuais: China (US $ 41bi); Brasil (US $ 18bi); Rússia (US $ 18bi); Índia (US $ 18bi) e África do Sul (US $ 5bi).
 

### 1.2.2. SAÚDE

A cooperação na área de saúde remonta à I Reunião de Ministros da Saúde do BRICS, em 2011, à qual se seguiram mais oito encontros. Essa vertente de cooperação tem resultado na identificação de problemas comuns, como a incidência de doenças infecciosas e a falta de acesso equitativo a serviços médicos e a medicamentos. Como realizações concretas da cooperação em saúde do BRICS, tem-se a criação da Rede de Pesquisa em Tuberculose, que visa a reunir esforços de pesquisa e desenvolvimento no combate a essa doença. No âmbito das discussões multilaterais de saúde, o BRICS reúne-se à margem da Assembleia Mundial da Saúde (AMS), desde 2012, e coordena-se na defesa da Declaração de Doha sobre o Acordo TRIPS e Saúde Pública, de 2001.  


### 1.2.3. CIÊNCIA, TECNOLOGIA & INOVAÇÃO

A vertente de cooperação em CT&I no BRICS tem-se revelado uma das mais promissoras. Iniciada em 2014, com a primeira reunião de ministros da área, tem produzido resultados palpáveis, tanto em termos de intercâmbio de conhecimento, quanto de recursos disponibilizados para projetos de pesquisa. Atualmente, essa frente de cooperação conta com onze grupos de trabalho, que tratam de temas que vão desde tecnologia geoespacial até gestão de recursos hídricos, passando por biotecnologia e biomedicina, infraestrutura de pesquisa e megaprojetos de ciência, tecnologia da informação e comunicação, entre outras áreas de interesse.

A cooperação BRICS em CT&I tem potencial para fomentar pesquisas que levem à produção de bens de alto valor tecnológico agregado; ao aumento da quantidade de patentes tanto nacionais quanto compartilhadas; à criação de rede de parques tecnológicos e ao intercâmbio de conhecimento entre os países. O principal objetivo dessa cooperação é gerar resultados econômicos concretos a partir da inovação.
 

### 1.2.4. SEGURANÇA

O diálogo sobre segurança do BRICS dá-se, sobretudo, nos encontros entre as autoridades nacionais de segurança dos cinco países – encontro de NSA – e nos grupos de trabalho da área de segurança. Por meio desse diálogo, os cinco países manifestam suas abordagens sobre as ameaças à segurança internacional e sobre o tratamento de crimes transnacionais, como tráfico de drogas, ataques cibernéticos, lavagem de dinheiro, corrupção e terrorismo. Em 2019, o encontro de NSA e as atividades dos grupos de trabalho avançarão sob a presidência de turno brasileira, durante a qual se discutirá o prosseguimento da cooperação já existente, bem como será priorizado o combate a ilícitos transnacionais.
 

### 1.2.5. EMPRESARIAL

O diálogo empresarial dá-se, sobretudo, por meio do Conselho Empresarial do BRICS (CEBRICS) e pelo Fórum Empresarial do BRICS.

 O CEBRICS foi estabelecido em 2013, durante a Cúpula de Durban, na África do Sul, com vistas a aproximar as comunidades empresariais dos cinco países, promover o compartilhamento de experiências e a prospecção de oportunidades de negócios. Atualmente, o CEBRICS já conta com nove grupos de trabalho, inclusive nas áreas de infraestrutura, manufatura, energia, agronegócio, serviços financeiros, aviação regional, harmonização de padrões técnicos e desenvolvimento de capacidades. Em 2018, estabeleceu-se novo grupo de trabalho sobre Economia Digital, dentro da estrutura do conselho. Durante o ano de 2019, a presidência de turno brasileira pretende avançar com a pauta de aproximação entre o CEBRICS e o NDB, a fim de impulsionar os investimentos em infraestrutura.¹

## 2.1. Formato dos dados
São cinco tabelas em csv com as seguintes colunas: *SeriesName, SeriesCode, CountryName, CountryCode, Year e Value*.

A coluna *SeriesName* determina o Indicador ou Indíce em Análise, exemplo de Indicador seria "Total tax and contribution rate (% of profit)" que informa o total de impostos pagos sobre o lucro das empresas. A coluna *SeriesCode* identifica o Indicador ou Indíce por um Código no formato "IC.TAX.TOTL.CP.ZS".

Já as colunas *CountryName* e *CountryCode* identificam o País a que se referem aqueles valores, *CountryName* informa o nome do País em inglês e *CountryCode* informa o Código daquele país, *vb. CountryName* = Brazil e *CountryCode* = BRA.

Por fim, temos a coluna *Year* e *Value* que informam, respectivamente, o Ano e o Valor daquele Indicador ou Indíce.

Também é importante salientar que conforme a necessidade os dados da coluna *Value* serão normalizados para que seja possível comparar proporcionalmente as grandezas, como por exemplo, proporcionalidade entre População e valor do Produto Interno Bruto, pois os valores podem, e, na maioria dos casos, terão uma gritante diferença de escala sendo necessário este tratamento nos dados.

Abaixo mostramos o *head()* ou primeiras cinco linhas de cada dataset para melhor entendimento.

```python
# formato dos dados
df_array = {}

for k in paths.keys():
    df_array[k] = pd.read_csv(str(paths[k]),sep=';')
    
for k in df_array.keys():
    display(k)
    display(df_array[k].head())
    df_array[k].info()
    
```

## 2.2. Pré-processamento e tratamento dos dados

<p style="text-align:justify">
Como podemos observar nas informações acima, os dados estão separados em 5 datasets, agrupados por país e SeriesName. 
</p>
<p style="text-align:justify">
    Podemos observar uma grande quantidade de nulos no campo "*Value*" das "*Series*" ou Indíces, tornando inviável a utilização de alguns indicadores em nosssa análise, sendo necessário descartar os dados que não podem ser utilizados.
</p>
<p style="text-align:justify">
    Sabendo que os dados estão agrupados por *SeriesName*, devemos analisar a viabilidade de cada série de forma independente. Neste trabalho consideraremos viáveis para análise as "*Series*" com menos de 30% de dados nulos.
</p>
<p style="text-align:justify">
    Abaixo iniciamos as tratativas necessárias para utilização dos dados nos gráficos, dessa forma criando uma nova base de dados que contém apenas as "*Series*" ou Indíces considerados "bons" o qual chamamos de "good_data_df".

```python
good_data_df = pd.DataFrame()
bad_data_df = pd.DataFrame()
total_data_df = pd.DataFrame()

for k in df_array.keys():
    
    # removo as 3 ultimas linhas de todos os datasets por serem dados não relevantes e não processaveis
    # (fonte de onde os dados foram obtidos, data de extração, etc)
    series = df_array[k].SeriesName.unique()
    series = series[0:len(series)-3]
    
    # para cada serie
    for s in series:
        # vejo qual o perncentual de dados nulos
        df_serie = df_array[k][df_array[k].SeriesName == s]
        per_null = (df_serie.Value.isna().sum()/len(df_serie.index))*100
        
        total_data_df = total_data_df.append(df_serie)
        
        # separo os dados com menos de 30% de nulos
        if(per_null < 30):
            good_data_df = good_data_df.append(df_serie)
        else:
            bad_data_df = bad_data_df.append(df_serie)
            
        

display("Series com menos de 30% de nulos")
display("Quantidade de series: "+str(len(good_data_df.SeriesName.unique())))
good_data_df.info()


display("Series com 30% ou mais de nulos")
display("Quantidade de series: "+str(len(bad_data_df.SeriesName.unique())))
bad_data_df.info()


display("Todos os dados")
display("Total de series: "+str(len(total_data_df.SeriesName.unique())))
total_data_df.info()
```

Como podemos ver nas saídas acima no quadro "Series com 30% ou mais de nulos" há um gap enorme de dados em muitas "*Series*", mais especificamente, num total de 1372 "*Series*", apenas 472 possui menos de 30% dos dados nulos, ou seja, são considerados viáveis para análise.

## 3.2. Riquezas

Em termos de geração de riqueza, qual o país mais rico? Será o mais populoso? É isso que vamos descobrir a seguir ao analisarmos os índices que medem a economia dos países. O principal índice que devemos avaliar é o GDP mais conhecido por nós como PIB que conforme a OECD é:

Original:
> Gross domestic product (GDP) is the standard measure of the value added created through the production of goods and services in a country during a certain period. As such, it also measures the income earned from that production, or the total amount spent on final goods and services (less imports). While GDP is the single most important indicator to capture economic activity, it falls short of providing a suitable measure of people's material well-being for which alternative indicators may be more appropriate. This indicator is based on nominal GDP (also called GDP at current prices or GDP in value) and is available in different measures: US dollars and US dollars per capita (current PPPs). All OECD countries compile their data according to the 2008 System of National Accounts (SNA). This indicator is less suited for comparisons over time, as developments are not only caused by real growth, but also by changes in prices and PPPs.

Traduzido (PT-BR):
> Produto interno bruto (PIB) é a medida padrão do valor adicionado criado pela produção de bens e serviços em um país durante um determinado período. Como tal, também mede a receita obtida com essa produção, ou o valor total gasto em bens e serviços finais (menos importações). Embora o PIB seja o indicador mais importante para capturar a atividade econômica, ele não fornece uma medida adequada do bem-estar material das pessoas, para a qual indicadores alternativos podem ser mais apropriados. Este indicador é baseado no PIB nominal (também chamado de PIB a preços correntes ou PIB em valor) e está disponível em diferentes medidas: dólares americanos e dólares americanos per capita (PPPs correntes). Todos os países da OCDE compilam seus dados de acordo com o Sistema de Contas Nacionais (SNA) de 2008. Este indicador é menos adequado para comparações ao longo do tempo, uma vez que a evolução não é causada apenas pelo crescimento real, mas também por alterações nos preços e PPPs.

Para nossa análise utilizaremos o GDP/PIB *per capita* ou "por cabeça" em dólares americanos de cada um dos cinco países em análise.

Apenas tendo em mente o tamanho populacional de cada país pode-se assumir erroneamente que os países mais populosos geram mais riquezas, logo seriam os mais ricos. Mas ao analisarmos o gráfico podemos observar que a população sozinha não tem praticamente nada ver com o quanto o PIB dividido por cabeça representa para cada país.

Na verdade, até mesmo a análise fria do PIB per capita não é capaz de indicar se há uma boa distribuição de renda no país, não conseguimos avaliar o bem-estar da população daquele país, pois apesar de gerar muita riqueza, se essa riqueza estiver extremamente concentrada em apenas algumas pessoas o bem-estar geral pode ser ruim ou péssimo.

Dessa forma, podemos concluir que a população em si não influencia a riqueza, observe a India o segundo país mais populoso ficando atrás apenas da China, tem o pior PIB per capita entre os países desde 1994, mantendo-se constante com um *gap* significativo em relação a África do Sul que é um país jovem e pequeno num continente com diversos problemas.

## 3.3. Conclusões

Para que possamos realizar esta comparação, necessitamos novamente tratar os dados, de forma que possamos colocá-los no mesmo gráfico, a razão disso reside no fato de que cada "*Series*" possui a sua informação em escalas de grandeza diferentes, o que tornaria a comparação direta inútil sem a normalização dos dados, pois as "*Series*" que possuirem uma menor escala terão um valor insignificante em comparação com as outras "*Series*", e por consequência o gráfico se tornaria ilegível.

Para normalização dos dados será utilizado a biblioteca de pré processamento do scikit-learn que possui todas as funções de cálculo necessárias para a normalização de dados.

```python
from sklearn import preprocessing

#normalizo os dados de uma das series

#separação de uma das series para normalização
df_gdp_good = good_data_df[good_data_df.SeriesName == 'GDP per capita (current US$)']

# criação de um objeto normalizador do scikit-learn
min_max_scaler = preprocessing.MinMaxScaler()

# aplicação da função de normalização na coluna de valores da serie 
df_gdp_normalizado =  min_max_scaler.fit_transform(df_gdp_good[['Value']])

# substituo os valores antigos da serie pelos valores normalizados
df_gdp_good['Value'] = df_gdp_normalizado


#normalizo os dados da outra serie
# e repito os passos do trecho de codigo acima
df_pop_good = good_data_df[good_data_df.SeriesName == 'Population, total']

min_max_scaler = preprocessing.MinMaxScaler()

df_pop_normalizado =  min_max_scaler.fit_transform(df_pop_good[['Value']])

df_pop_good['Value'] = df_pop_normalizado


# faço uniao das duas series de dados
df_final = df_gdp_good.append(df_pop_good)

fiapig = px.line(df_final, 
              x="Year", 
              y="Value", 
              color='CountryName', 
              line_group="SeriesName", 
              hover_name="SeriesName",
              title='PIB per capita / populacao total',
              labels={'Year':'Ano','Value':'Valor'}).update_traces(connectgaps=True)
fiapig.show()
```

No gráfico acima vemos que os países de forma isolada, em sua maioria, quanto maior a população, maior o PIB per capita, todavia para comprovarmos se isso é mesmo correto, veremos a correlação destes dados utilizando uma matriz de correlação relacionando os dados da população e do PIB em seus respectivos países e anos. 

```python
# separo o dataframe sem o indice original 
# para poder colocar as colunas de valores das series juntas
df_1 = df_gdp_good.reset_index()

# pego o valor e dou nome significativo
df_1["PIB"] = df_1["Value"]

# separo o dataframe sem o indice original 
# para poder colocar as colunas de valores das series juntas
df_2 = df_pop_good.reset_index()
# pego o valor e dou nome significativo
df_2["População"] = df_2["Value"]

# concateno a coluna de valores da populacao pelo eixo de coluna
# sem o indice original dos datasets a ordenação fica correta por data e país
df_correlacao = pd.concat([df_1,df_2["População"]],axis=1)

# elimino antiga coluna com nome value e coluna do indice antigo 
df_correlacao = df_correlacao.drop(["Value","index"],axis=1)

# montagem da matriz de correlação
z = df_correlacao.corr()
labels = list(z.columns)

colorscale = [[0,'red'],[0.5, 'white'], [1, 'green']]
fig = ff.create_annotated_heatmap(np.array(np.round(z,2)), colorscale=colorscale, x=labels, y=labels)
fig.update_layout(title_text='PIB per capita / populacao total')
fig.show()
```

Conforme a matriz de correlação, podemos ver que na verdade a relação entre PIB per capita e população é, na verdade, um pouco inversa. Podemos perceber isso analisando o comportamento da China com o Brasil, que apesar de os dois paises de forma isolada, com o passar dos anos, aumentar sua população e seu PIB, quando comparados um com o outro percebemos que o Brasil com uma menor populacao, possui um PIB maior, ou seja, tem maior produção de riquezas com uma menor população, demonstrando assim esta proporção inversa do PIB em relação à população apresentada pela matriz de correlação.
