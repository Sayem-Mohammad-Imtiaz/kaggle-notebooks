### Correlation Between Numerical Values
#### Understanding Correlation
Correlation shows the strength of a relationship between two variables and is expressed numerically by the correlation coefficient. The correlation coefficient's values range between -1.0 and 1.0. A perfect positive correlation means that the correlation coefficient is exactly 1. This implies that as one security moves, either up or down, the other security moves in lockstep, in the same direction. A perfect negative correlation (-1.0) means that two assets move in opposite directions, while a zero correlation implies no linear relationship at all.


For example, large-cap mutual funds generally have a high positive correlation to the Standard and Poor's (S&P) 500 Index or nearly one. Small-cap stocks have a positive correlation to the S&P, but it's not as high or approximately 0.8.

However, put option prices and their underlying stock prices will tend to have a negative correlation. For review, a put option gives the owner the right, but not the obligation, to sell a specific amount of an underlying security at a pre-determined price within a specified time frame. Put option contracts become more profitable when the underlying stock price decreases. In other words, as the stock price increases, the put option prices go down, which is a direct and high-magnitude negative correlation.

Calculated correlation (r) between two variables value:
* r<0.20 and values close to zero are not related or indicates a very weak relationship.
* Weak correlation between 0.20-0.39
* Moderate correlation between 0.40-0.59
* A high level of correlation between 0.60-0.79
* If it is 0.80-1.0, it is interpreted as a very high correlation.

```python
list1 = ["Customer_care_calls","Customer_rating","Cost_of_the_Product","Prior_purchases","Discount_offered","Weight_in_gms","Reached.on.Time_Y.N"]
plt.figure(figsize = (16,5))
sns.heatmap(data[list1].corr(), annot = True, fmt = ".2f")
plt.show()
```

As you can see in the correlation chart above, the most remarkable relationships: 
* Reached on Time - Discount Offered (0.40)
* Customer Care Calls - Cost of The Product (0.32)
* Weight of The Product in Grams - Discount Offered (-0.38)

### Relation Between Product Arrived on Time - Discount Offered

```python
data1=data[data["Reached.on.Time_Y.N"]==1].sort_values("Discount_offered",ascending=True)
a=list(Counter(data1["Discount_offered"]).keys())
b=list(Counter(data1["Discount_offered"]).values())


data2=data[data["Reached.on.Time_Y.N"]==0].sort_values("Discount_offered",ascending=True)
c=list(Counter(data2["Discount_offered"]).keys())
d=list(Counter(data2["Discount_offered"]).values())

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=a,
    y=b,
    name = '<b>not</b> reached on time (1)', # Style name/legend entry with html tags
))
fig.add_trace(go.Scatter(
    x=c,
    y=d,
    name='reached on time (0)',
))

fig.update_layout(
    title="Discount and Delay Relationship",
    xaxis_title="Discount Offered",
    yaxis_title="Number of Discounted Products",
    font=dict(
        family="Courier New, monospace",
        size=13,
        color="RebeccaPurple"
    )
)
fig.show()
```

As can be seen in the graphic above, the products that received more than 10 US dollars discount did not arrive on time. 10 dollars and less discounted items were more often delivered to the buyer on time. So, if you've ordered a product with a discount of more than 10 dollars, don't expect it to arrive on time.

### Relation Between Customer Care Calls - Cost of The Product

```python
fig = px.density_heatmap(data, x="Customer_care_calls", y="Cost_of_the_Product", nbinsx=10, nbinsy=10, color_continuous_scale="Viridis")


fig.show()
```

As seen above, the prices of the most ordered products are between 250 and 300 in 6 or 7 calls to customer service, while the prices of the products in 4 or 5 calls are between 200 and 250 at most. For 2 or 4 calls, the price drops to between 150-200 currency. In this case, as the price of the product gets more expensive, people tend to call customer service more. Of course, it would be much more preferable to give customer service a headache instead of losing an order worth 300 dollars.

### Relation Between Weight of The Product in Grams - Discount Offered

```python
f, ax = plt.subplots(figsize=(15, 8))
sns.despine(f, left=True, bottom=True)
clarity_ranking = ["I1", "SI2", "SI1", "VS2", "VS1", "VVS2", "VVS1", "IF"]
sns.scatterplot(x="Discount_offered", y="Weight_in_gms",
                palette="ch:r=-.2,d=.3_r",
                hue_order=clarity_ranking,
                sizes=(1, 8), linewidth=0,
                data=data, ax=ax)
```

According to this graph, most of the discounts applied to products are less than 10 dollars, and products weighing more than 4 kilos are not discounted more than 10 dollars, with some exceptions.

### Which Mode of Shipment is Faster Than Others?

```python
labels = ["Products Arriving on Time","Products Not Arriving on Time"]

fig = make_subplots(1,3, specs=[[{'type':'domain'}, {'type':'domain'},{'type':'domain'}]],
                    subplot_titles=['Ship', 'Flight',"Road"])
fig.add_trace(go.Pie(labels=labels, values=[3003 , 4459], scalegroup='one',
                     name="Ship"), 1, 1)
fig.add_trace(go.Pie(labels=labels, values=[708 , 1069], scalegroup='one',
                     name="Flight"), 1, 2)
fig.add_trace(go.Pie(labels=labels, values=[725 , 1035], scalegroup='one',
                     name="Road"), 1, 3)

fig.update_layout(title_text='Which Mode of Shipment is Faster Than Others?')
fig.show()
```

According to the chart above, orders were mostly transported by ships. The number of materials transported by road and air is almost the same. The delivery rates of the products to the buyer on time are considered the same in 3 ways. Unfortunately, you will not be able to tell company officials which routa must be used your product to arrive faster.

### What is The Number of Times The Products Stored In Each Warehouse Arrive on Time?

```python
labels = ["Products Arriving on Time","Products Not Arriving on Time"]

fig = make_subplots(2,3, specs=[[{'type':'domain'}, {'type':'domain'},{'type':'domain'}],[{'type':'domain'}, {'type':'domain'},{'type':'domain'}]],
                    subplot_titles=["A","B","C","D","F"])
fig.add_trace(go.Pie(labels=labels, values=[758 , 1075], scalegroup='one',
                     name="A"), 1, 1)
fig.add_trace(go.Pie(labels=labels, values=[729 , 1104], scalegroup='one',
                     name="B"), 1, 2)
fig.add_trace(go.Pie(labels=labels, values=[739 , 1094], scalegroup='one',
                     name="C"), 1, 3)
fig.add_trace(go.Pie(labels=labels, values=[738 , 1096], scalegroup='one',
                     name="D"), 2, 1)
fig.add_trace(go.Pie(labels=labels, values=[1472 , 2194], scalegroup='one',
                     name="F"), 2, 2)

fig.update_layout(title_text='What is The Number of Times The Products Stored In Each Warehouse Arrive on Time?')
fig.show()
```

As you will see, the timely arrival of your order does not depend on the warehouse where it is stored or the way it arrives. However, we can say that most of the products are stored in the F warehouse.

### Which Gender is More Satisfied With The Orders?

```python
ratings=["1","2","3","4","5"]
fig = go.Figure()
fig.add_trace(go.Bar(
    x=ratings,
    y=[1086, 1103, 1096, 1093, 1076],
    name='Male Rating',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=ratings,
    y=[1149, 1062, 1143, 1096, 1095],
    name='Female Rating',
    marker_color='lightsalmon'
))

# Here we modify the tickangle of the xaxis, resulting in rotated labels.
fig.update_layout(barmode='group',title_text="Rating For Each Genders")
fig.show()
```

According to the graph, the rating ratios of the genders are almost the same. Therefore, we can say that gender does not matter in rating.

### Which products were the most liked by ways of transportation?

```python
labels = ["1","2","3","4","5"]

fig = make_subplots(1,3, specs=[[{'type':'domain'}, {'type':'domain'},{'type':'domain'}]],
                    subplot_titles=['Ship', 'Flight',"Road"])
fig.add_trace(go.Pie(labels=labels, values=[1505, 1456, 1566, 1469, 1466], scalegroup='one',
                     name="Ship"), 1, 1)
fig.add_trace(go.Pie(labels=labels, values=[369, 363, 362, 353, 330], scalegroup='one',
                     name="Flight"), 1, 2)
fig.add_trace(go.Pie(labels=labels, values=[361, 347, 343, 357, 352], scalegroup='one',
                     name="Road"), 1, 3)

fig.update_layout(title_text='Which products were the most liked by ways of transportation?')
fig.show()
```

As you will see, just like the genders, the transportation route of the products has no effect on the ratings.
