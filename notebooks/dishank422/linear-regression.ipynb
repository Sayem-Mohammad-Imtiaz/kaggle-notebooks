```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
happiness2015_data = pd.read_csv("../input/world-happiness/2015.csv", index_col = 0)
happiness2016_data = pd.read_csv("../input/world-happiness/2016.csv", index_col = 0)
happiness2015_data.head()
```

```python
x1 = np.array([happiness2015_data['Economy (GDP per Capita)'], happiness2015_data['Family'], happiness2015_data['Health (Life Expectancy)'], happiness2015_data['Freedom'], happiness2015_data['Trust (Government Corruption)'], happiness2015_data['Generosity'], happiness2015_data['Dystopia Residual']])
x2 = np.array([happiness2016_data['Economy (GDP per Capita)'], happiness2016_data['Family'], happiness2016_data['Health (Life Expectancy)'], happiness2016_data['Freedom'], happiness2016_data['Trust (Government Corruption)'], happiness2016_data['Generosity'], happiness2016_data['Dystopia Residual']])
x1_final = np.transpose(x1)
x2_final = np.transpose(x2)
training_data_features = np.concatenate((x1_final, x2_final))
np.shape(training_data_features)
```

```python
y1 = np.array([happiness2015_data['Happiness Score']])
y2 = np.array([happiness2016_data['Happiness Score']])
training_data_dependent_variable = np.transpose(np.concatenate((y1, y2), axis = 1))
```

```python
np.shape(training_data_dependent_variable)
```

```python
model = LinearRegression().fit(training_data_features, training_data_dependent_variable)
model.coef_
```

```python
model.score(training_data_features, training_data_dependent_variable)
```
