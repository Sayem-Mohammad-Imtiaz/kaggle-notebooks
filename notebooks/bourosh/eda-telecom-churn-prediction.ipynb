Жители какого штата говорят по телефону больше всего? меньше всего? Обратите внимание, что нас интересуют все виды звонков.

Как связаны наличие голосовой почты и международного плана с лояльностью абонентов? Проверьте статистическую значимость этих связей.

```python
from scipy.stats import fisher_exact
fisher_exact(pd.crosstab(df['international plan'], df['churn']))
```

```python
from scipy.stats import chi2_contingency
chi2_contingency(pd.crosstab(df['international plan'], df['churn']))
```

Какие из числовых признаков имеют наибольшую степень корреляции? Почему?

```python
numeric = [ 'total day calls', 'total day charge', 'total eve charge', 'total eve minutes', 'total eve calls', 'account length', 'number vmail messages', 'total day minutes', 'total night minutes', 'total night calls', 'total night charge', 'total intl minutes', 'total intl calls', 'total intl charge', 'customer service calls']
sns.pairplot(df[numeric]);
```

Признаки количества звонков за какое-либо время и оплата клиента за какое-либо время имеют наибольшую степень корреляции.

Связано ли время пользования аккаунтом с количеством обращений в колл-центр?

```python
from scipy.stats import pearsonr
pers = pearsonr(df['customer service calls'], df['account length'])
print('Pearson correlation:', pers[0], 'p-value:', pers[1])
```

```python
from scipy.stats import spearmanr
spearm = spearmanr(df['account length'], df['customer service calls'])
print('Spearman correlation:', spearm[0], 'p-value:', spearm[1])
```

```python
from scipy.stats import kendalltau
kendall = kendalltau(df['account length'], df['customer service calls'])
print('Kendall correlation:', kendall[0], 'p-value:', kendall[1])
```

Коэффициенты корреляций очень малы (почти 0), то есть  взаимосвязь между временем пользования аккаунтом и количеством обращений в колл-центр практически незначима.
