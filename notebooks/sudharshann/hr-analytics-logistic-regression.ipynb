**Now,we'll choose our required attributes from dataset to work with.**

```python
Req_Col = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
Req_Col.head()
```

**Now,we create dummies for high,low and medium salary.**

```python
dummy = pd.get_dummies(Req_Col.salary,prefix='salary')

# Now we attach that dummy to the final df

final = pd.concat([Req_Col,dummy],axis='columns')

#Now we can drop the salary attribute as we've created dummies for them
final.drop('salary',axis = 'columns',inplace = True)
final.head(3)
```

# Now,we're assigning x and y

```python
X = final
y = df.left
X.head()
```

# **Now, our dataset is all set to perform operations.**

 **Here we'll be using logistic regression to classify**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size=0.40,random_state = 1)
```

# Now we're fitting the model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

# ** We got 78.5 percent accuracy**
