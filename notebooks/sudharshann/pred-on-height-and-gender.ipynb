# converting all categorical values to numerical values.

```python
df['Gender'].replace('Male',0,inplace = True)
df['Gender'].replace('Female',1,inplace = True)
df.head(),df.tail()
```

# Choosing Required attributes

Now,we'll choose our required attributes from dataset to work with.

```python
Req_Col = df[['Gender','Height']]
Req_Col.head()
```

Assigning X and y

Now,we're assigning x and y

```python
X = Req_Col
y = df.Weight
X.head(), y.head()
```

# Splitting dataset into training set and test set.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.14, random_state=1)
```

Here we're giving 14 percentage of our data to test and 86 to train

# Fitting the model

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

# Checking accuracy

```python
model.score(X_test,y_test)
```

**We've 91 percent accuracy**
