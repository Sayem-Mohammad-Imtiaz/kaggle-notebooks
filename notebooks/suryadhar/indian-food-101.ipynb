```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns 
```

#                                         loading the Data set from the kaggle

```python
ds=pd.read_csv('/kaggle/input/indian-food-101/indian_food.csv')
```

```python
ds.isnull().sum()
```

We have one null value at the region . Let explore and impute with some meaningfull in sightsdata. The region was missing for the Uttar Pradesh state. As we know the UP belongs to the North. We can replace the Null value with the North  

```python
ds.at[110,'region']='North'
```

```python
ds.iloc[110]['region']
```

```python
ds.isnull().sum()
```

```python
ds.describe()
```

 In Preap Time and Cook Time we have -1 as the values which is not possible so let us consider the same as the Null values. Therefore, replacing the -1 with the Null Values 

```python
ds.replace(-1,np.nan,inplace=True)
```

```python
ds.isnull().sum()
```

Total 30 and 28 null values in the Prep_time and Cook_time. Before imputing Null values with mean/meadian or mode let us check the distribution of the values. The Distribution of the histogram is rightly sckewed so imputing the Values with the median

```python
ds.prep_time.plot(kind='hist')
```

```python
ds.prep_time.median()
```

```python
ds.prep_time.fillna(ds.prep_time.median(),inplace =True)
```

```python
ds.cook_time.plot(kind='hist')
```

```python
ds.cook_time.median()
```

```python
ds.cook_time.fillna(ds.cook_time.median(),inplace=True)
```

```python
ds.cook_time.isnull().sum()
```

```python
ds.cook_time.plot(kind='hist')
```

After exploring the data we have also found that there are values -1 in flavor_profile, state and region. Hence we have replaced the string -1 wit NaN values.

```python
ds.replace('-1',np.nan,inplace=True)
```

```python
ds.isnull().sum()
```

Impted the Null(NaN) Values so that the -1 will be excluded from  our analysis 

```python
l=ds.flavor_profile
sns.set_style('darkgrid')
plt.figure(figsize=(10,10))
sns.countplot(x='flavor_profile',data=ds,order=ds.flavor_profile.value_counts().index.tolist())
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.xlabel('Flavor_profile',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.title('Flavour Count Plot',fontsize=14,weight='bold')
```

                  In India Most of the People prefer the Spicy Food followed by the Sweet

```python
l=text.split(',')

len(l)
```

```python
from collections import Counter
l=Counter(l)

```

To Convert the list into the dict based on their frequancy we have used the Counter from collections

```python
ing=pd.DataFrame(data=l,index=(range(491)))
ingredients=ing.T[0].reset_index()
ingredients.rename(columns={'index':'ingredients',0:'No of times'},inplace=True)
P=ingredients.sort_values('No of times',ascending=False)[0:15]
plt.figure(figsize=(10,10))
sns.barplot(y='ingredients',x='No of times',data=P)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('No of Times',fontsize=12)
plt.ylabel('ingredients',fontsize=12)
plt.title('Most Used Ingredients',weight='bold',fontsize=14)
```

Same Visulization as **Bar graph**
