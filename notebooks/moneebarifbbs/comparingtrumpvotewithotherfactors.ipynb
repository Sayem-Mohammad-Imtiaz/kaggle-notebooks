```python
import numpy as num
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

> Loading Data from cvs and visualizing it

```python
elections = pd.read_csv('/kaggle/input/2020-general-election-polls/county_statistics.csv')
elections.head(5)
```

Now visulizing all the columns.

```python
print(elections.dtypes);
```

Now cleaning data. 
1) i want to study change in prenagte of votes trumps with other factors so new colunms 'differentInTrumpPrecntage'
2) Men and Women votes in percentage colunms.
3) Converte death , cases and votingcitizen into precentage.


```python
elections['differentInTrumpPrecntage'] = elections['percentage20_Donald_Trump']- elections['percentage16_Donald_Trump'] 
elections['differentInTotalVotes'] = elections['total_votes20']- elections['total_votes16'] 
elections['womeninprecntage'] = 100*elections['Women']/elections['TotalPop'] 
elections['meninprecntage'] = 100*elections['Men']/elections['TotalPop'] 
elections['deathsinPrecntage']= 100 *elections['deaths']/elections['TotalPop'];
elections['casesinPrecntage']= 100*elections['cases']/elections['TotalPop'];
elections['VotingAgeinPrecntage']= 100*elections['VotingAgeCitizen']/elections['TotalPop'];
```

1) removing *unused varables like lat, long , and all the columns for which we convert it to prentage and opposition vote count*

```python
elections = elections.drop(['percentage20_Joe_Biden', 'percentage16_Hillary_Clinton', 'percentage20_Donald_Trump', 'percentage16_Donald_Trump','total_votes16',
  'votes16_Donald_Trump', 'votes20_Donald_Trump' , 'votes16_Hillary_Clinton' , 'votes20_Joe_Biden','cases', 'deaths', 'lat', 'long', 'Men','Women', 'VotingAgeCitizen','TotalPop', 'state','county'],axis=1)

elections.head()
```

Seeing colums left behind


```python
print(elections.dtypes);
```

Droping NA

```python
print(elections.isnull().any().any());
elections=elections.dropna()
```

Ploting heatmap

```python
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(elections.corr()[['differentInTrumpPrecntage']].sort_values('differentInTrumpPrecntage').tail(100),
 vmax=1, vmin=-1, cmap='YlGnBu', annot=True, ax=ax);
ax.invert_yaxis()
```

  From this heatmap most important factor which increase vote of trump is people in construction and Hispanic.
  And also look like most wealthy county change the vote most against trump.
