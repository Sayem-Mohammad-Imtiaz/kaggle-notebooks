# This data is analyzing the floating population in South Korea during the Covid crisis.
## Definition of a floating population: A group of people who reside in a given population for a certain amount of time and for various reasons, but are not generally considered part of the official census count

### Import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
import re
import datetime
import folium
import geopandas as gpd
from shapely.geometry import Point, Polygon
from geopandas import GeoDataFrame
import math
from folium.plugins import HeatMap, MarkerCluster
from folium import Marker,GeoJson,Choropleth, Circle
```

### Import data and view first 5 rows

```python
floating = pd.read_csv('/kaggle/input/coronavirusdataset/SeoulFloating.csv')
 
```

```python
floating.head()
```

## Transform the genders to integers

```python
le = LabelEncoder()
floating.sex = le.fit_transform(floating.sex)
```

## I cannot plot the 'hour', 'bith_year' and 'sex' columns without transforming them into numeric objects, which I did with the method below

```python
floating.birth_year = pd.to_numeric(floating.birth_year, errors = 'coerce')
floating.sex = pd.to_numeric(floating.sex, errors = 'coerce')
floating.hour = pd.to_numeric(floating.hour, errors = 'coerce')
```

## The distribution of the floating numbers

```python
sns.distplot(floating.fp_num)
```

## The distribution of the sexes in the floating numbers show there are more females than males

```python
sns.violinplot('sex','fp_num', data = floating )
```

## Because the same city is represented at least twice to caputure the different sexes, hour of the day and so one, I grouped the data by the city to capture the entire population in each city

```python
city = floating.groupby(['city'], as_index = False).sum()
```

```python
city
```

```python
plt.figure(figsize = (8, 6))
sns.barplot('fp_num','city', data = city)
```

## To see the population by date, I grouped by date and then visualized it

```python
date = floating.groupby('date', as_index = False).sum()
```

```python
date
```

```python
# On any given day the population is fairly constant

plt.figure(figsize = (20,20))
sns.scatterplot('fp_num','date', data = date)
```

## Now I want to see how the population looks like in each given hour, so I grouped the data by hour and then did a line plot

```python
hour = floating.groupby('hour', as_index = False).sum()
```

```python
hour.head()
```

```python
# So in a given day this is the distribution of floating population in the cities

sns.lineplot('hour', 'fp_num', data = hour)
```

## The distribution of the population according to birth year is also done by grouping
