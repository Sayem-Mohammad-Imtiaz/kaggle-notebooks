```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder
import re
import datetime
import folium
import geopandas as gpd
from shapely.geometry import Point, Polygon
from geopandas import GeoDataFrame
import math
from folium.plugins import HeatMap, MarkerCluster
from folium import Marker,GeoJson,Choropleth, Circle
```

## Import data and view first few rows

```python
policy = pd.read_csv('/kaggle/input/coronavirusdataset/Policy.csv')  
```

```python
policy.head(10)
```

## Making the 'state_date' and 'end_date' numeric, so I can subtract the two dates to get the duration of the different policies

```python
policy.start_date = pd.to_datetime(policy.start_date, errors = 'coerce')
policy.end_date = pd.to_datetime(policy.end_date, errors = 'coerce')
```

```python
policy['duration'] = (policy['end_date'] - policy['start_date'])
```

```python
policy.head(10)
```

## Removing the string words from the duration

```python
policy['duration'] = policy['duration']/np.timedelta64(1, 'D')
```

```python
policy.head()
```

## The distribution of the duration

```python
sns.distplot(policy.duration)
```

## A barplot of the distribution of the duration with the types of policies show Education lasted the longest
### Any blank section is because of the null present in the data

```python
sns.barplot( 'duration', 'type',data = policy)
```

## Those for School reopenings with online class lasted the longest

```python
plt.figure(figsize = (7,5))
sns.barplot( 'duration', 'gov_policy',data = policy)
```

## This gives more detail of the types of policies with their durations

```python
plt.figure(figsize = (15,13))
sns.barplot( 'duration', 'detail',data = policy)
```
