```python
df_gpby = df[['parental level of education','race/ethnicity','TotalScore']]
grouped_test1 = df_gpby.groupby(['parental level of education','race/ethnicity'],as_index=False).mean()
grouped_test1
```

The above table is not easy to visualize or to make any inference out of it. Let's make the above table in pivot with assigning the parameter to specific index and columns.

```python
grouped_pivot = grouped_test1.pivot(index='race/ethnicity', columns='parental level of education')
grouped_pivot
```

Now, the above table is easy to read and tells us what's actually going on when we group the race/ethnicity and the parental level of education. As per the above, results it looks like as we go from group A to group E the number tends to increase in the total score of the students in all the levels of degree. To get more idea about it let's map this into a heat map to get more colours and have better understanding of the correlation.

```python
plt.pcolor(grouped_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(grouped_pivot, cmap='RdBu')

#label names
row_labels = grouped_pivot.columns.levels[1]
col_labels = grouped_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

The above heat maps cleary shows how the scores varies between the groups and the parental level of education. The heap map states that students whose parents are with Master's degree scored well as compared to the other level of education. Then comes the students whose parents have the Bachelors's degree.

Creating dummy variables for all the categorical variables.

```python
df['genderdummy'] = np.where(df.gender=='female',0,1)
df['lunchdummy'] = np.where(df.lunch == 'standard',1,0)
df['coursedummy'] = np.where(df['test preparation course']=='none',0,1)
```

```python
df.head()
```

```python
assigndegree = {"bachelor's degree":3, 'some college':4, "master's degree":5,
       "associate's degree":6, 'high school':1, 'some high school':2}
df['degreedummy'] = df['parental level of education'].map(assigndegree)
```

```python
assignrace = {'group B':2, 'group C':3, 'group A':1, 'group D':4, 'group E':5}
df['racedummy'] = df['race/ethnicity'].map(assignrace)
```

```python
df.head()
```

Model Development

```python
from sklearn.linear_model import LinearRegression
```

```python
z = df[['racedummy','degreedummy', 'genderdummy', 'lunchdummy' , 'coursedummy']]
lm = LinearRegression()
lm.fit(z, df['TotalScore'])
lm.intercept_
```

```python
lm.coef_
```

Testing and Training Model

```python
from sklearn.model_selection import train_test_split
X = df.drop(['TotalScore', 'parental level of education', 'race/ethnicity','test preparation course',
                   'lunch', 'gender', 'math score', 'reading score', 'writing score'], axis = 1)
Y = df['TotalScore']
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
print("number of test samples :", x_test.shape[0])
print("number of training samples:",x_train.shape[0])
```

```python
x_train.head()
```

```python
from sklearn.linear_model import LinearRegression
lm1= LinearRegression()
lm1.fit(x_train,y_train)
lm1.score(x_test,y_test)
```
