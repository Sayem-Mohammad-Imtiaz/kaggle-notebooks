### Data Visualization

```python
#Bar Chart of Math Scores
df['math score'].value_counts().head(5).plot.bar(color=['#eb4034','#ebd334','#ffba93','#349ceb','#ae34eb'])
```

```python
#Bar Chart of Reading Scores
df['reading score'].value_counts().head(5).plot.bar(color=['#eb4034','#ebd334','#ffba93','#349ceb','#ae34eb'])
```

```python
#Bar Chart of Writing Scores
df['writing score'].value_counts().head(5).plot.bar(color=['#eb4034','#ebd334','#ffba93','#349ceb','#ae34eb'])
```

```python
df['parental level of education'].value_counts().plot.bar(color=['#eb4034','#ebd334','#ffba93','#349ceb','#ae34eb','#be383b'])
```

```python
#Line Chart
df['math score'].value_counts().sort_index().plot.line()
```

```python
#Area Chart
df['math score'].value_counts().sort_index().head(50).plot.area() #For first 50 Data
```

```python
#Histogram
df.plot.hist()
```

```python
df['math score'].plot.hist()
```

```python
df[df['math score'] < 50]['math score'].plot.hist()
```

```python
#Scatter Plot
df[df['reading score'] < 50].plot.scatter(x='reading score', y='math score',color='#A42008')
```

```python
#Hex Plot
df[df['math score'] < 80].plot.hexbin(x='math score', y='reading score',gridsize=10) 
```

```python
#Stacked Bar Chart
df_score2=df_scores.head(30)
df_score2.plot.bar(stacked=True)
```
