Lets create a heat graph of the data set we are give in order to observe some correlations. For example, does the popularity of a director in anyway influence the score of a film?

```python
f=pd.read_csv("../input/movie_metadata.csv")
```

```python
f.corr()

```

```python
correlation = f.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation, vmax=1, square=True,annot=True,cmap='cubehelix')

plt.title('Correlation in Movie Data Set')
```

We can deduce a number of things from this heat map. The strongest correlation we notice in the data set is between the number of critics who have reviewed the film and the number of facebook likes for the film (0.68). Obviously this means that the more popular a film is the more it will reviewed (or vice versa).
Unfortunately to answer our primary question, the popularity of the director does not seem to influence the public's perception of the film in any way deemed significant as the correlation resides at a small 0.17.
