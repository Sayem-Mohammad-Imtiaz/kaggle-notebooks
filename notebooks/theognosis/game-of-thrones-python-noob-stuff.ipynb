Has the number of character deaths increased by each book?

```python
f=pd.read_csv("../input/character-deaths.csv")

```

```python
Book1 = 0
Book2 = 0
Book3 = 0
Book4 = 0
Book5 = 0
deaths_bybook=f["Book of Death"]
for deaths in deaths_bybook:
    if deaths == 1.0:
         Book1 = Book1 + 1
print(Book1)

deaths_bybook=f["Book of Death"]
for deaths in deaths_bybook:
    if deaths == 2.0:
         Book2 = Book2 + 1
print(Book2)

deaths_bybook=f["Book of Death"]
for deaths in deaths_bybook:
    if deaths == 3.0:
         Book3 = Book3 + 1
print(Book3)

deaths_bybook=f["Book of Death"]
for deaths in deaths_bybook:
    if deaths == 4.0:
         Book4 = Book4 + 1
print(Book4)

deaths_bybook=f["Book of Death"]
for deaths in deaths_bybook:
    if deaths == 5.0:
         Book5 = Book5 + 1
print(Book5)
    
    
```


    We got the data we were looking for but lets display it on a plot!

```python
Allbooks = [Book1, Book2, Book3, Book4, Book5]
plt.plot(Allbooks)
```

It seems that George Martin's addiction for death and misery was somewhat appeased by the end of the third book as the number of characters sacrificed by the author have gone considerably down by the start of the 4th. Yet it seems to increase once again in the fifth. Let us pray for our god's compassion for the books to come...(or you can pray for his cruelty, whatever is most entertaining really)

```python
labels = 'Got', 'CoK', 'SoS', 'FfC', 'DwD'
sizes = [49, 73, 97, 27, 61]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red']
explode = (0, 0, 0.1, 0, 0)  # explode  slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```

```python
Alldeaths = Book1 + Book2 + Book3 + Book4 + Book5
print(Alldeaths)
```

As we can see, GRR Martin killed a total of 307 characters throughout the Song of Ice and Fire series. Thats a lot! We also found out that the number of deaths in the 3rd book count for 31,6% of all the deaths in total. Thats almost a third of the series. The third book is often considered the most popular of the series while the fourth book, the one with the least amount of deaths in the series, is often considered the worst.
