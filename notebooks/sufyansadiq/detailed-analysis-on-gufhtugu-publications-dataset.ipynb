**Handling missing values**

**Number of missing values in each column**

```python
missing_data=gp_orders.isnull()
for column in missing_data.columns.values.tolist():
    print(column)
    print(missing_data[column].value_counts())
    print("")
```

**Replacing missing values in "Payment method" column, with most occuring value in that column**

```python
print(gp_orders["Payment Method"].value_counts())
frequent0="Cash on delivery"
```

```python
gp_orders["Payment Method"].replace(np.NaN, frequent0 ,inplace=True)
```

**Replacing missing values in "City" column, with most occuring value in that column**

```python
print(gp_orders["City"].value_counts())
frequent1="Karachi"
```

```python
gp_orders["City"].replace(np.NaN, frequent1, inplace=True)
```

Replacing missing values in "Book Name" column, with most occuring value in that column

```python
print(gp_orders["Book Name"].value_counts())
frequent2="Python Programming"
```

```python
gp_orders["Book Name"].replace(np.NaN,frequent2,inplace=True)
```
