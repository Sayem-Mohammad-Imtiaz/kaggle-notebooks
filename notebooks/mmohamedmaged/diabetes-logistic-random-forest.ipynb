There are many columns that have zeros value almost zeroes values presents missing values

```python
X=diabetes.copy()
```

```python
for i in ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']:
        print('Num of missing values in {0} is {1}'.format(i,sum(X[i]==0)))
```

```python
sns.countplot(x=X['Outcome'])        
```

```python
X.hist(bins=40 , figsize=(20,20)) 
plt.show()
```

remember in skinThichness ,Glaucose , insulin,bloodpressure,bmi zeroes represent missing 
According to these distributions we can fill BMI , Glucose , SkinThickness and BloodPressure  with mean

```python
X[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']]=X[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.nan)
```

```python
for i in ['Glucose','BloodPressure','SkinThickness','BMI']:
                X[i].fillna(X[i].mean(), inplace = True)
X['Insulin'].fillna(X['Insulin'].median(), inplace = True)
```

```python
sns.relplot(x='Glucose',y='Insulin',data=X,hue='Outcome')
```

```python
sns.relplot(x='BMI',y='Age',data=X,hue='Outcome')
```

```python
msno.bar(X)
```

There is no missing now

```python
X.loc[X['Age']<=30 ,'Age']=1
X.loc[(X['Age']>30) & (X['Age']<45),'Age']=2
X.loc[X['Age']>=45 ,'Age']=3
```

```python
sns.relplot(x='BMI',y='Age',data=X,hue='Outcome')
```

```python
sns.countplot(x=X['Age'],hue=X['Outcome'])        
```

```python
p=scatter_matrix(X,figsize=(25, 25))
```

```python
#[1--> under weight , 2--> Normal , 3-->over,4--> above]
X.loc[X['BMI']<=18.5 ,'BMI']=1
X.loc[(X['BMI']>18.5) & (X['BMI']<=24.9),'BMI']=2
X.loc[(X['BMI']>24.9) & (X['BMI']<29.9),'BMI']=3
X.loc[X['BMI']>=29.9 ,'BMI']=4
```

```python
#Standardize the data
for i in ['BloodPressure','SkinThickness','Insulin']:
            X[i]=(X[i] - np.mean(X[i])) /np.std(X[i])
```

```python
plt.figure(figsize=(12,10))  
p=sns.heatmap(X.corr(), annot=True)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X.iloc[:,:-1], X.iloc[:,-1],test_size=1/2,random_state=42)
```

```python
LogisticRegressionModel = LogisticRegression()
LogisticRegressionModel.fit(X_train, y_train)

#Calculating Details
print('LogisticRegressionModel Train Score is : ' , LogisticRegressionModel.score(X_train, y_train))
print('LogisticRegressionModel Test Score is : ' , LogisticRegressionModel.score(X_test, y_test))
```

```python

RandomForestClassifierModel = RandomForestClassifier(criterion = 'gini',n_estimators=19,max_depth=3,random_state=33) 
RandomForestClassifierModel.fit(X_train, y_train)

#Calculating Details
print('RandomForestClassifierModel Train Score is : ' , RandomForestClassifierModel.score(X_train, y_train))
print('RandomForestClassifierModel Test Score is : ' , RandomForestClassifierModel.score(X_test, y_test))
```
