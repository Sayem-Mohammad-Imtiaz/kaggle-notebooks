# Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

To understand the data visually, let's define functions. One for countrywise visualization, one for regionwise visualization. 

```python
def visualizer_by_country (aspect, number_of_countries):
    grouped_country=countrywise.groupby('Country/Region')[str(aspect)]
    grouped_country.sum().sort_values(ascending=False)[:number_of_countries].plot(kind='bar')
    return grouped_country
```

```python
visualizer_by_country(aspect='Deaths', number_of_countries=5);
```

```python
visualizer_by_country(aspect='Deaths / 100 Cases', number_of_countries=5);
```

```python
visualizer_by_country(aspect='Confirmed', number_of_countries=5);
```

```python
def visualizer_by_region (aspect, number_of_countries):
    grouped_country=countrywise.groupby('WHO Region')[str(aspect)]
    grouped_country.sum().sort_values(ascending=False)[:number_of_countries].plot(kind='bar', color='r')
    return grouped_country
```

```python
visualizer_by_region (aspect='Deaths', number_of_countries=5);
```

```python
visualizer_by_region (aspect='Confirmed', number_of_countries=5);
```

```python
visualizer_by_region (aspect='Deaths / 100 Cases', number_of_countries=5);
```

Death rate seems to be the highest in 'Yemen' countrywise and Europe in regionwise.

```python
clean_complete=pd.read_csv('../input/corona-virus-report/covid_19_clean_complete.csv')
```

```python
clean_complete.head()
```

Let's analyze the countries and regions having the most death month by month. To do that let's define functions. Again one for countryvise visualization and one for regionwise visualization. In functions we will define a month column and plot according to that so we can reduce the samples to be plotted.

```python
def visualize_month_by_month_country(country,aspect):
    df=clean_complete.drop(['Lat','Long','Recovered','Active','Province/State'],axis=1)
    df2=df[df['Country/Region']==str(country)]
    df2=df2.reset_index()
    df2=df2.drop('index',axis=1)
    df2.Date=pd.to_datetime(df2.Date)
    df2['Month']=df2.Date.dt.month
    sns.barplot(x=df2.Month,y=df2[str(aspect)],color='gold')
    plt.xticks(ticks=[0,1,2,3,4,5,6,7],labels=['January','February','March','April','May','June','July','August'], rotation=90)
    plt.show()
```

```python
visualize_month_by_month_country(country='US',aspect='Deaths')
```

```python
visualize_month_by_month_country(country='Brazil',aspect='Deaths')
```

```python
visualize_month_by_month_country(country='India',aspect='Deaths')
```

```python
visualize_month_by_month_country(country='Russia',aspect='Deaths')
```

```python
def visualize_month_by_month_region(region,aspect):
    df=clean_complete.drop(['Lat','Long','Recovered','Active','Province/State'],axis=1)
    df2=df[df['WHO Region']==str(region)]
    df2=df2.reset_index()
    df2=df2.drop('index',axis=1)
    df2.Date=pd.to_datetime(df2.Date)
    df2['Month']=df2.Date.dt.month
    sns.barplot(x=df2.Month,y=df2[str(aspect)],color='darkorange')
    plt.xticks(ticks=[0,1,2,3,4,5,6,7],labels=['January','February','March','April','May','June','July','August'], rotation=90)
    plt.show()
```

```python
visualize_month_by_month_region(region='Americas',aspect='Deaths')
```

```python
visualize_month_by_month_region(region='Europe',aspect='Deaths')
```

Deaths and illness had increased with time consistently. This is expected. Nothing interesting happening there.

There are 58 states to visualize. That's a big number. To plot successfully, we will determine a limit and reject the ones below this limit. Let's define a function which can do this for us and construct a pie chart. 

```python
US_grouped['Percentage']=US_grouped.Deaths.apply(lambda x: x*100/sum(US_grouped.Deaths))
```

```python
def pie_creator_for_bigger_data(df,name_col,percentage_col,percentage_limit,width,height):
    df2=df[df[percentage_col]>percentage_limit]
    fig=px.pie(df2,values=percentage_col,names=name_col, width=width,height=height)
    return fig
```

```python
pie_creator_for_bigger_data(df=US_grouped,
                             name_col='Province_State',
                             percentage_col='Percentage',
                             percentage_limit=1.0,
                             width=1000,
                             height=1000)
```
