# Create a global map for COVID Vaccinations(Percentage of the population that are fully vaccinated).

I choose the daily vaccinated number and change the color scale. Since some data has exceed 100000per day, so I set the color range relatively large to cover all the data.

```python
import numpy as np
import pandas as pd 
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
#print(df.columns)
#print(df['daily_vaccinations'])
fig = px.choropleth(df,locations='country',locationmode='country names',color='daily_vaccinations',hover_name='country',animation_frame='date',color_continuous_scale=px.colors.sequential.Viridis,range_color=(0.0,1000000.0))
fig.update_layout(title_text="Daily vaccinations in each country",title_x=0.45)
fig.show()

plotly.offline.plot(fig,filename='daily_vaccinations.html')
```

# Create a global map of earthquake occurrences worldwide.

```python
df = pd.read_csv('../input/202105-recen-earthquake/query.csv',header=0)
#print(df.columns)
df.index = pd.to_datetime(df['time'])
df['time'] = df.index.strftime('%Y-%m-%d %H:00:00')
df = df.sort_index()
fig = px.scatter_geo(df,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='rainbow',range_color=(1.0,3.0))
#print(df['time'])
fig.show()
```
