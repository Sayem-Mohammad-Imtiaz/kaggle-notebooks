```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import re
from tqdm.notebook import tqdm
import nltk

import chart_studio.plotly as py
import plotly.graph_objs as go
import plotly.figure_factory as ff
import plotly.express as px
from plotly.subplots import make_subplots
from plotly.offline import iplot
from wordcloud import WordCloud
from plotly.offline import iplot

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_csv("../input/all-trumps-twitter-insults-20152021/trump_insult_tweets_2014_to_2021.csv")[['target', 'tweet']]
data.head()
```

<h2 align='center' style='color:green'>Character Frequency Count!</h2>

```python
char_freq = data['tweet'].str.len()
fig = px.histogram(x=char_freq)
fig.update_layout(title="Character Frequency Count")
iplot(fig)
```

<h2 align='center' style='color:green'>Word Count Distribution!</h2>

```python
word_dist = data['tweet'].str.split().map(lambda x: len(x))
fig = px.histogram(x=word_dist, color_discrete_sequence=px.colors.sequential.Agsunset)
fig.update_layout(title="Word Count Distribution")
iplot(fig)
```

<h2 align='center' style='color:green'>Unique Word Count</h2>

```python
uwc = data['tweet'].apply(lambda x: len(set(str(x).split()))).to_list()
fig = ff.create_distplot([uwc], ['All Tweets'])
fig.update_layout(title="Unique Word Count in all Tweets")
iplot(fig)
```

<h2 align='center' style='color:green'>URL Count</h2>

```python
url = data['tweet'].str.split().map(lambda x: len([w for w in str(x).lower().split() if 'http' in w or 'https' in w or 'ftp' in w]))
fig = px.histogram(x=url, color_discrete_sequence=px.colors.sequential.algae_r)
fig.update_layout(title="URL Count")
iplot(fig)
```

<h2 align='center' style='color:green'>Word Cloud</h2>

```python
wc = " ".join(data['tweet'].tolist())
wc = WordCloud(width=256, height=256, collocations=False).generate(wc)
plt.figure(figsize=(8, 8))
plt.imshow(wc)
plt.axis('off')
plt.show()
```

<h2 align='center' style='color:green'>Bi-Gram Words</h2>

```python
data['bigrams'] = data['tweet'].apply(lambda x: list(nltk.bigrams(nltk.word_tokenize(x))))
data['bigrams_count'] = data['bigrams'].apply(lambda x: len(x))

data.head()
```

```python
plt.style.use('classic')
sns.displot(data['bigrams_count'], color='red', kde=True)
plt.title("Bigrams Count Distribution")
plt.xlabel("Number of Bigrams")
plt.ylabel("Count")
plt.show()
```
