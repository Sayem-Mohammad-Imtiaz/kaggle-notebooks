```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
df = pd.read_csv('../input/so-survey-2017/survey_results_public.csv')
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
print('Total survey responses: ', df.shape[0])
```

```python
df['Country'].value_counts()[:10]
```

```python
sns.set(style = 'whitegrid')
plt.figure(figsize = (15,7))
plt.title('Survey responses per country')
ax = sns.countplot(x = 'Country', data = df, color = 'darkgoldenrod', order = df['Country'].value_counts()[:10].index)

for p in ax.patches:
        ax.annotate('{:.1f}'.format(p.get_height()) , (p.get_x()+0.1, p.get_height()+50))
        #((what you want annotated), (x position, y position))
```

```python
df['Professional'].values
```

```python
student = df[df['Professional'] == 'Student'] # makes it so we only have student responses

plt.figure(figsize = (15, 7))
ax = sns.countplot(x = 'Country', data = student, order = student['Country'].value_counts()[:10].index, color = 'sandybrown')

for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x() + 0.25, p.get_height() + 0.55))
```

```python
plt.figure(figsize = (15,8))
plt.xticks(rotation = 90)
sns.countplot(x = 'Professional', data = df, order = df['Professional'].value_counts().index)
```

```python
profdev = df[df['Professional'] == 'Professional developer']
profdev.shape
```

```python
plt.figure(figsize = (18,8))
ax = sns.countplot(x = 'Country', data = profdev, order = profdev['Country'].value_counts()[:10].index)

for p in ax.patches:
    ax.annotate(format(p.get_height()), (p.get_x() + 0.25, p.get_height() + 0.55))
```

```python
d={}
for i in ['C','C++','C#','Java','Python','R']:
    d[i]=(df['HaveWorkedLanguage'].apply(lambda x: i in str(x).split('; ')).value_counts()[1])
    
lang=pd.DataFrame(list(d.items()))
lang.columns=[['Language','Count']]
lang.set_index('Language',inplace=True)    
lang.plot.barh(width=0.8,color='#005574')
plt.title('Popularity of each language')
plt.show()
```
