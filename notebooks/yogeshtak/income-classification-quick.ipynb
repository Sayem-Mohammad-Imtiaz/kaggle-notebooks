**Let's just make a quick model**

```python
data['income'] = data['income'].map({" <=50K": 0, " >50K": 1})
```

```python
data.head()
```

**Target**
1. Income less than or equal to 50K : 0
2. Income greater than 50K : 1

```python
#split X and y
X = data.drop('income',axis=1)
y = data['income']


#conveting categorical data into numerical data
import category_encoders as ce
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ['workclass','education','marital-status','occupation',
                        'relationship','race','sex','native-country']
one_hot = ce.OneHotEncoder()
transformer = ColumnTransformer([("one_hot", one_hot, categorical_features)],
                                remainder = "passthrough") #pass through columns not specified in categorical features

transformed_X = transformer.fit_transform(X)
```

```python
transformed_X = pd.DataFrame(transformed_X)
transformed_X.head()
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split

np.random.seed(23)

X_train, X_test, y_train, y_test = train_test_split(transformed_X,y, test_size=0.2)

model = LinearSVC()
model.fit(X_train,y_train)

model.score(X_test,y_test)
```
