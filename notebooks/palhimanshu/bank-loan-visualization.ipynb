```python
data=pd.read_csv('../input/my-dataset/credit_train.csv')
```

```python
#import pygal
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data.head()
```

```python
data.info()
```

***As we knows that,before providing loan bank gather some information depends on the Annual Income of the person and also the account history.So that according to the Annual Income removing null values. ***

```python
#dropping the null values
data.dropna(inplace=True,subset=['Annual Income','Credit Score'])
```

```python
#filling some of the null values
data.fillna(method='bfill',inplace=True)
```

```python
data.describe()
```

```python
plt.figure(figsize=(15,9))
sns.countplot(x='Purpose',data=data,hue='Loan Status')
plt.xticks(rotation=45)
plt.title('Purpose of taking Loan',fontdict={'fontsize':20})
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(x=data['Term'].value_counts(),labels=['short term','long term'],explode=[0,0.09],)
plt.title('Time Period of Taking Loan',fontdict={'fontsize':20})
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x='Home Ownership',data=data,hue='Loan Status')
plt.title('Own Property vs Loan Status',fontdict={'fontsize':20})
plt.show()
```

```python
data['Number of Open Accounts']=data['Number of Open Accounts'].astype(np.int64)
plt.figure(figsize=(15,8))
sns.countplot(x=data['Number of Open Accounts'],hue=data['Bankruptcies'])
plt.title('Open Accounts vs Bankruptcies',fontdict={'fontsize':20})
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=data['Tax Liens'],y=data['Monthly Debt'])
plt.title('Dept vs Tax liens',fontdict={'fontsize':20})
plt.show()
```

*A federal tax lien is the government's legal claim against your property when you neglect or fail to pay a tax debt. The lien protects the government's interest in all your property, including real estate, personal property and financial assets.*

```python
plt.figure(figsize=(15,8))
sns.barplot(x=data['Bankruptcies'],y=data['Monthly Debt'],palette='dark')
plt.title('Bankruptcy vs Monthly Dept',fontdict={'fontsize':20})
plt.show()
```

***Bankruptcy is a legal proceeding involving a person or business that is unable to repay their outstanding debts. ... All of the debtor's assets are measured and evaluated, and the assets may be used to repay a portion of outstanding debt.***
