```python
#first of all importing libraries
import seaborn as sns
import matplotlib.pyplot as plt
import plotly
```

```python
#reading the data
data=pd.read_csv('../input/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv')
```

```python
data.head()
```

```python
#Information Regarding the data
data.info()
```

```python
#no duplicate values
data.duplicated()
```

```python
#renaming some ofthe  Column names
data.rename(columns={'State/UTs':'State','Active Ratio (%)':'Active ratio','Discharge Ratio (%)':'Discharge ratio','Death Ratio (%)':'Death ratio'},inplace=True)
```

```python
#renamed data
data.head()
```

# Performing Exploratory Data Analysis

```python
#here we use bar plot which is showing the Total case
plt.figure(figsize=(13,15))
sns.barplot(y='State',x='Total Cases',data=data)
plt.title('State-wise Total Cases',{'fontsize':20})
plt.show()
```

```python
#This plot is for the Active cases in the states
plt.figure(figsize=(13,15))
sns.barplot(y='State',x='Active',data=data,palette='Set2')
plt.title('State-wise Active Cases',{'fontsize':20})
plt.show()
```

```python
#Recovery from the covid-19
plt.figure(figsize=(15,8))
sns.barplot(x='State',y='Discharged',data=data,palette='husl')
plt.xticks(rotation=80)
plt.title('State-wise Recovery',{'fontsize':20})
plt.show()
```

```python
# Total Deaths state-wise
plt.figure(figsize=(13,15))
sns.barplot(y='State',x='Deaths',data=data,palette='husl')
plt.title('State-wise Deaths',{'fontsize':20})
plt.show()
```

```python
#this pie plot is showing the Active ratio in the different states
plt.figure(figsize=(15,15))
plt.pie(data['Active ratio'],labels=data['State'],rotatelabels=True,labeldistance=1)
plt.title('Active Ratio',{'fontsize':20})
plt.show()
```

```python
#Recovery ratio
plt.figure(figsize=(15,15))
plt.pie(data['Discharge ratio'],labels=data['State'],rotatelabels=True,labeldistance=1)
plt.title('Recovery Ratio',{'fontsize':20})
plt.show()
```

```python
#Death ratio
plt.figure(figsize=(15,15))
plt.pie(data['Death ratio'],labels=data['State'],rotatelabels=True,labeldistance=1)
plt.title('Death Ratio',{'fontsize':20})
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.plot(data['Deaths'])
plt.plot(data['Active'])
plt.legend(['Deaths','Active'])
plt.title('Deaths VS Active',{'fontsize':20})
plt.axis('off')
plt.show()
```
