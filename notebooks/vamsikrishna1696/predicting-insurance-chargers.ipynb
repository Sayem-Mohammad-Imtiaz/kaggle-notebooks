```python
#importing data
data=pd.read_csv('/kaggle/input/insurance/insurance.csv')
```

```python
#Retreiving top 5 rows
data.head(5)
```

```python
#we need to check for null/NaA values as models wont accept we need to remove them or fill them
data.isnull().sum()
```

```python
#checking DataTypes
data.dtypes
```

```python
#We selected object data Type
data.select_dtypes(object)
```

```python
# As models only accept numerical data type we need to convert object data types to numerical. with labelencoder we can convert object data type to integer
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
#converting object datatypes to integers
data['sex']=le.fit_transform(data['sex'])
data['smoker']=le.fit_transform(data['smoker'])
data['region']=le.fit_transform(data['region'])
```

```python
# checking data types now we have all integers so we can feed them to the model
data.dtypes
```

```python
#Regression model requires Independent and Depandant variables 
x=data.iloc[:,0:6]
```

```python
#independent variable
y=data['charges']
```

```python
#splitting the dataset into train and test
from sklearn.model_selection import train_test_split
trainx,testx,trainy,testy=train_test_split(x,y, test_size=0.2)
```

```python
#LinearRegression Model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#training regression model with training data
model.fit(trainx,trainy)
```

```python
#In the above step model has already learnt with the training data now we are predicting the independent variable with test data
ypred=model.predict(testx)
```

```python
#coefficients obtainted from model
model.coef_
```

```python
#intercept
model.intercept_
```

```python
#comparing Actual values vs predicted values
pd.DataFrame({'Actual': testy, 'Predict': ypred})
```

```python
#comparing actual vs predicted with scatterplot
import matplotlib.pyplot as plt
plt.scatter(testy,ypred, color='red')
```
