```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/kaggle/input/breast-cancer-wisconsin-data/data.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
df['Unnamed: 32'].isna().sum() == len(df)
```

```python
df = df.drop(columns = ['Unnamed: 32'])
```

```python
labels = df['diagnosis']
```

```python
df_no_labels = df.drop(columns = ['diagnosis'])
```

```python
df_no_labels = df_no_labels.drop(columns = ['id'])
```

```python
ss = StandardScaler()
df_no_labels_scaled = pd.DataFrame(ss.fit_transform(df_no_labels), columns = df_no_labels.columns)
```

```python
df_no_labels_scaled
```

```python
labels[labels == 'M'] = 0
```

```python
labels[labels == 'B'] = 1
```

```python
labels = labels.astype(int)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_no_labels_scaled, labels, test_size = 0.2, random_state = 42)
```

```python
y_train.head()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
logis = LogisticRegression()
fores = RandomForestClassifier()
```

```python
y_train.head()
```

```python
logis.fit(X_train, y_train)
fores.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred_logis = logis.predict(X_test)
y_pred_fores = fores.predict(X_test)
a1 = accuracy_score(y_pred_logis, y_test)
a2 = accuracy_score(y_pred_fores, y_test)
print(a1,a2)
```
