```python
df = pd.read_csv('../input/suicide-rates-overview-1985-to-2016/master.csv')

df
```

```python
#show unique countries and age-groups 
countries = np.unique(df['country'])
Age_group = np.unique(df['age'])
print('Countries :')
print(countries)
print('Age Groups :')
print(Age_group)
```

```python
#visualize the age group who is in higher risk of suicide
united_states = df.where(df['country'] == 'United States')
united_states = united_states.where(df['year'] >=2000)
united_states = united_states.dropna()
fig = plt.figure()
plt.subplot(221)
ax = fig.add_axes([0,0,1,1])
age_group = united_states['age']
suicide_number = united_states['suicides/100k pop']
ax.bar(age_group, suicide_number)
plt.title('Suicide age group in US')

United_Kingdom = df.where(df['country'] == 'United Kingdom')
United_Kingdom = United_Kingdom.where(df['year'] >=2000)
United_Kingdom = United_Kingdom.dropna()
fig = plt.figure()
plt.subplot(222)
ax = fig.add_axes([0,0,1,1])
age_group = United_Kingdom['age']
suicide_number = United_Kingdom['suicides/100k pop']
ax.bar(age_group, suicide_number)
plt.title('Suicide age group in UK')
```
