Importing the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
#1. Reading the train data
data_train = pd.read_csv('../input/health-insurance-cross-sell-prediction/train.csv')
```

```python
data_train.head()
```

```python
data_train.info()
```

```python
data_train.describe()
```

```python
data_train["Region_Code"].unique()
```

```python
#2. Cleaning data
data_train.isnull().sum()
# No null value in the dataset
```

```python
#3. Processing data
#gender
gender_map = {'Female':1, 'Male':2}
data_train['Gender'] = data_train['Gender'].map(gender_map)
data_train.head()
```

```python
data_train["Vehicle_Age"].unique()
```

```python
#Vehicle_Age
vehicle_age_map = {'< 1 Year':1, '1-2 Year':2, '> 2 Years':3}
data_train['Vehicle_Age'] = data_train['Vehicle_Age'].map(vehicle_age_map)
data_train.head()
```

```python
data_train["Vehicle_Damage"].unique()
```

```python
#Vehicle_Damage
vehicle_damage_map = {'Yes':1, 'No':2}
data_train['Vehicle_Damage'] = data_train['Vehicle_Damage'].map(vehicle_damage_map)
data_train.head()
```

```python
#4.Determining x and y
x=data_train[["Gender", "Age", "Driving_License", "Region_Code","Previously_Insured", "Vehicle_Age", "Vehicle_Damage", "Annual_Premium", "Policy_Sales_Channel", "Vintage" ]]
x.head()
```

```python
y=data_train[["Response"]]
y.head()
```

```python
#5. splitting data into train and test
x_train, x_test, y_train, y_test=train_test_split(x, y, random_state=6)
```

```python
#standardizing x data
transfer = StandardScaler()
x_train = transfer.fit_transform(x_train)
x_test = transfer.transform(x_test)
```

```python
#6. Using different machine learning methods to test the forcast accuracy
#Using Random Forest
estimator=RandomForestClassifier(n_estimators=10, criterion="entropy", max_depth=8, bootstrap=True, max_features="auto")
estimator.fit(x_train, y_train)
y_predict=estimator.predict(x_test)
accuracy = estimator.score(x_test, y_test)
print("The accuracy by RandomForest:\n", accuracy)
```

```python
#Using KNN
estimator = KNeighborsClassifier(n_neighbors=3)
estimator.fit(x_train,y_train)
y_predict=estimator.predict(x_test)
accuracy = estimator.score(x_test, y_test)
print(accuracy)
```

```python
#Using Xgboost
estimator= XGBClassifier()
estimator.fit(x_train, y_train)
y_predict=estimator.predict(x_test)
accuracy = estimator.score(x_test, y_test)
print("The accuracy by Xgboost:\n", accuracy)
```

```python
#Using Logistic Regression
estimator = LogisticRegression()
estimator.fit(x_train, y_train)
y_predict=estimator.predict(x_test)
accuracy = estimator.score(x_test, y_test)
print(accuracy)
report=classification_report(y_test, y_predict, labels=[0, 1])
print(report)
```

Logistic Regression shows a high accuracy. We will use Logistic Regression to forcast the responses in test dataset.
