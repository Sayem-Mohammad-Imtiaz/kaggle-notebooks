# 3. age

```python
df[df['age'].isnull()].count()
```

```python
df['age'].groupby(df['race']).mean()
```

```python
def impude_age (cols):
    age = cols[0]
    race = cols[1]
    if pd.isnull(age):
        if race == 'A':
            return 37
        elif race == 'B':
            return 32
        elif race == 'H':
            return 34
        elif race == 'N':
            return 31
        elif race == 'O':
            return 33
        else:
            return 40
    else:
        return age
        
```

```python
df['age']=df[['age','race']].apply(impude_age, axis=1)
```

```python
sns.heatmap(df.isnull())
```

# 4. race
