# PLOTTING PIE CHART FOR TYPE (MOVIE and TV-SHOWS)

```python
import seaborn as sns
import matplotlib.pyplot as plt



ax = plt.subplots(figsize = (10,10))
plt.pie(x=data["type"].value_counts(),labels=data["type"].value_counts().index , autopct= "%.2f%%" )
plt.title('Distribution of Movie/TV Show', size=15)
```

# FINDING THE MOST COMMON RATINGS

```python
plt.figure(figsize=(15,6))
temp= data.rating.value_counts()
sns.barplot(x="index" , y="rating" , data = temp.reset_index())
plt.xlabel("RATING", size=12)
plt.ylabel("COUNT", size=12)
plt.title("TREND OF RATING", size=24)
```

# PLOTTING THE TOP 5 RATINGS 

```python
ax = plt.subplots(figsize = (8,8))
sns.barplot(x = data['rating'].value_counts().index[0:5], y = data['rating'].value_counts()[0:5])
```

# RELEASE OVER LAST 5 YEARS

```python
ax = plt.subplots(figsize = (10,10))
sns.barplot(x = data['release_year'].value_counts().index[:5], y = data['release_year'].value_counts()[:5])
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Release Frequency For Last 5 Years')
```

```python
data.head(2)
```

# TYPE VS COUNT FOR LAST 5 YEARS

```python
movie_data = data[data['type'] == 'Movie']
tv_show_data = data[data['type'] == 'TV Show']
# bar,ax = plt.subplots(1,2,figsize=(10,10))
temp = data[['type', 'release_year']]
temp = temp.value_counts().to_frame()
temp.reset_index(level=[0,1],inplace=True)
temp=temp.rename(columns={0:'count'})
temp = pd.concat([temp[temp['type'] == 'Movie'][:5], temp[temp['type']== 'TV Show'][:5]]) # for last 5 release_years
temp
```

```python
sns.catplot(x = 'release_year', y = 'count', hue = 'type', data = temp, kind = 'point')
plt.rcParams['figure.figsize']=(10,10)
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.title('Growth of Movie/TV Show over Years', size=12)
```

*Here we can see that trend for movies is constanlty decreasing from 2017*

*Whereas TV Shows are gaining popularity in the recent 5 years*
