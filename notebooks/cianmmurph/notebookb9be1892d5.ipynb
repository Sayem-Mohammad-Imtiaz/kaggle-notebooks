```python
country_stats = vaccine_country.groupby(['country','date'])[['total_vaccinations_per_hundred','people_fully_vaccinated_per_hundred', 'daily_vaccinations_per_million']].min()
```

```python
country_stats.dropna(inplace = True)
```

```python
country_stats.sort_index(level = 'date',ascending = False)
```

```python
neighbours = country_stats.loc[['Switzerland', 'Austria', 'France', 'Germany', 'Italy', 'Liechtenstein']]
```

```python
neighbours.sort_index(level = 'date', ascending = False)
neighbours.reset_index(inplace = True)
```

```python
neighbours.sort_values('date', ascending = False, inplace = True)
neighbours.reset_index(drop = True, inplace = True)
```

Here are the most recent statisitics for Switzerland and it's neighbouring countries

```python
neighbours.drop_duplicates(subset = 'country', keep = 'first',inplace = True)
neighbours.reset_index(drop = True, inplace = True)
neighbours
```

The following figure shows the number of vaccines administered per 100 people

```python
fig = plt.figure(figsize=(16,9))
fig.set_facecolor('white')
for i in range(0, len(neighbours)):
    plt.subplot(int(len(neighbours)/3), 3, i + 1)
    #prep plot
    to_plot = [neighbours.iloc[i][2],100-neighbours.iloc[i][2]]
    plot_labels = ["total vaccinations per hundred","remainder"] 
    #plot
    plt.pie(to_plot, labels=plot_labels, labeldistance=1.1, autopct='%1.1f%%')
    plt.title(neighbours.iloc[i][0])
    p=plt.gcf()
    
```

The following figure shows the number of people fully vaccinated per 100 people

```python
fig = plt.figure(figsize=(16,9))
fig.set_facecolor('white')
for i in range(0, len(neighbours)):
    plt.subplot(int(len(neighbours)/3), 3, i + 1)
    #prep plot
    to_plot = [neighbours.iloc[i][3], 100-neighbours.iloc[i][3]]
    plot_labels = ["people fully vaccinated per hundred","remainder"] 
    #plot
    plt.pie(to_plot, labels=plot_labels, labeldistance=1.1, autopct='%1.1f%%')
    plt.title(neighbours.iloc[i][0])
    p=plt.gcf()
    
```
