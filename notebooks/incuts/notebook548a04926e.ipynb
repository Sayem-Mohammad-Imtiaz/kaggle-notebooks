```python
import matplotlib.pyplot as plt
import seaborn as sns
```

**First Barplot**

```python
plt.figure(figsize=(18,4))
sns.barplot(x = "sepal_length", y = "species", hue = "species", data = df_iris, palette='pastel')
plt.show()
plt.figure(figsize=(18,4))
sns.barplot(x = "sepal_width", y = "species", hue = "species", data = df_iris, palette='pastel')
plt.show()
plt.figure(figsize=(18,4))
sns.barplot(x = "petal_length", y = "species", hue = "species", data = df_iris, palette='pastel')
plt.show()
plt.figure(figsize=(18,4))
sns.barplot(x = "petal_width", y = "species", hue = "species", data = df_iris, palette='pastel')
plt.show()
```

**Count plot**

```python
plt.figure(figsize=(23,5))
sns.countplot(x=df_iris['sepal_length'],data=df_iris,hue = 'species',palette='pastel')
plt.show()
plt.figure(figsize=(23,5))
sns.countplot(x=df_iris['sepal_width'],data=df_iris,hue = 'species',palette='pastel')
plt.show()
plt.figure(figsize=(23,5))
sns.countplot(x=df_iris['petal_length'],data=df_iris,hue = 'species',palette='pastel')
plt.show()
plt.figure(figsize=(23,5))
sns.countplot(x=df_iris['petal_width'],data=df_iris,hue = 'species',palette='pastel')
plt.show()
```

**Histogram**

```python
df_iris.hist(bins=50, figsize=(20,10))
plt.show()
```

**Pie plot**

```python
df_iris.groupby('species').mean().values
```

**The labels must be written in order: setosa -> versicolor -> virginica  ,, since they are grouped by species based on the dataset we have**

```python
labels = 'Iris-setosa', 'Iris-versicolor', 'Iris-virginica'
sizes = df_iris.groupby('species')['sepal_length'].mean().values
colors = ['lightblue', 'lightpink', 'lightgreen']
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Reading sepals length for every specie')
plt.show()

labels = 'Iris-setosa', 'Iris-versicolor', 'Iris-virginica'
sizes = df_iris.groupby('species')['sepal_width'].mean().values
colors = ['lightblue', 'lightpink', 'lightgreen']
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Reading sepals width for every specie')
plt.show()

labels = 'Iris-setosa', 'Iris-versicolor', 'Iris-virginica'
sizes = df_iris.groupby('species')['petal_length'].mean().values
colors = ['lightblue', 'lightpink', 'lightgreen']
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Reading petal length for every specie')
plt.show()

labels = 'Iris-setosa', 'Iris-versicolor', 'Iris-virginica'
sizes = df_iris.groupby('species')['petal_width'].mean().values
colors = ['lightblue', 'lightpink', 'lightgreen']
plt.pie(sizes, labels=labels, colors=colors,autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Reading petal width for every specie')
plt.show()
```

**Scatter plot**

```python
sns.set_style("whitegrid")
sns.pairplot(df_iris,hue="species",height=3);
plt.show()
```

```python
corr_matrix = df_iris.corr()
corr_matrix["petal_width"].sort_values(ascending=False)
```

**As shown here, petal length and sepal length have strong correlation to the petal width, they 'expel' with the petal width, whereas sepal width is 'opposite' correlated**

```python
corr_matrix = df_iris.corr()
plt.subplots(figsize=(20, 10))
sns.heatmap(corr_matrix, vmax=.9, square=True)
```
