## Competition Summary

```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go
from plotly.offline import iplot

import cufflinks
cufflinks.go_offline()
cufflinks.set_config_file(world_readable=True)
```

```python
PATH = '../input/chai-time-data-science/Results.csv'
results = pd.read_csv(PATH)
```

```python
#Copied from an Amazing Kernel by Anshur: https://www.kaggle.com/anshuls235/journey-of-ctds-show-through-visuals

fig = go.Figure()

fig.add_trace(go.Indicator(
    title = {'text':'Total Submissions Received','font':{'color': 'white'}},
    mode = "number",
    number={'font':{'color': 'white'}},
    value = results.Author.count(),
    domain = {'row': 0, 'column': 0}))


fig.add_trace(go.Indicator(
    title = {'text':'Total Unique Entries','font':{'color': 'white'}},
    mode = "number",
    number={'font':{'color': 'white'}},
    value = results.Author.nunique(),
    domain = {'row': 0, 'column': 1}))

fig.add_trace(go.Indicator(
    title = {'text':'Highest Points Scored','font':{'color': 'white'}},
    mode = "number",
    number={'font':{'color': 'white'}},
    value = results.Total.max(),
    domain = {'row': 1, 'column': 0}))

fig.add_trace(go.Indicator(
    title = {'text':'Average Total Points','font':{'color': 'white'}},
    mode = "number",
    number={'font':{'color': 'white'}},
    value = results.Total.mean(),
    domain = {'row': 1, 'column': 1}))


fig.update_layout(width=700,height=400,
                  title = {'text':'<b>CTDS.Show Contest Stats</b>',
                           'font':{'size':30,'color': '#0062ff'}},
                  template='seaborn',margin=dict(t=60,b=10,l=10,r=10),
                  grid = {'rows': 2, 'columns': 2, 'pattern': "independent"},paper_bgcolor='#07d7f2')
```

## Context about the Competition and Results:

The CTDS.Show Contest was run in an effort to celebrate the 1-year anniversary of the podcast and to invite insights from the Kaggle community for the podcast and things that may be improved. This was based on [Grandmaster Rohan Rao](https://www.kaggle.com/rohanrao)'s idea. [Read More about it here](https://www.kaggle.com/rohanrao/chai-time-data-science/discussion/168310#936323). 


After some Chai and brainstorming, the idea of hosting a competition was born, to make the "stars" (actually dates, but that was more dramatic), align, the timelines chosen were:

- Launching the competition on the date when the **first recording** was done.
- In true spirit of the podcast, hosting a few events to encourage participation by all levels of participants.
- Again, Kaggle Hereos and legend Kernel Grandmasters agreed to being judges for the competition!
- Finally, [Parul Pandey](https://www.kaggle.com/parulpandey) and [John Miller](https://www.kaggle.com/jpmiller) shared two amazing kernels on [Getting Started](https://www.kaggle.com/parulpandey/how-to-explore-the-ctds-show-data) and [Some Best Practices for Analytics Reporting](https://www.kaggle.com/jpmiller/some-best-practices-for-analytics-reporting) advices respectively

## Overall Leaderboard:

Based on the total points achieved, lets look at the absolute ranks (Higher is better)

```python
results = results.drop_duplicates(subset='Author', keep="first")

fig_overall = go.Figure()

fig_overall = px.bar(results.drop_duplicates(subset='Author', keep="first").sort_values(by=['Total']), y = "Author", x = "Total",
             title="Overall Leaderboard",
             hover_name = 'Author',
             text='Total',
             orientation='h', 
             width=650, height=1000, range_x = [0, 100])

fig_overall.update_traces(marker_color='#ff9900', textposition="outside")

fig_overall.show()
```

## Top 10: Breakdown of points achieved:

As you can see, the winner leads by a big gap and the gap shrinks towards the middle ranks, similar to a lot of Kaggle competitions :) 

```python
results = results.sort_values(by=['Total'],ascending=True)

fig_2 = go.Figure()

for col in results.columns[2:7]:
    fig_2.add_trace(go.Bar(y=results.Author, x=results[col],name=col, orientation='h'))
fig_2.update_layout(barmode='stack',width=650, height=1000)
fig_2.show()
```

## Categorical Winners:

To recap the categories were set as follows:

- **Presentation**: How well is the notebook written in terms of code quality, text description and grammar?
- **Storytelling**: Is there a natural flow of the story that connects various points?
- **Visualizations**: Are the visualizations appealing, understandable and aligned with description?
- **Insights**: Are the insights relevant, useful and actionable?
- **Innovation**: How novel and creative are the ideas and approaches?
