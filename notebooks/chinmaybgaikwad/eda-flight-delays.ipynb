```python
flights = pd.read_csv('../input/flight-delays/flights.csv')
flights
```

```python
flights_row, flights_cols = flights.shape
```

```python
Mis_Vals_Fl = flights.isnull().sum().to_frame('Missing_Values')
Mis_Vals_Fl['%_Missing'] = Mis_Vals_Fl.Missing_Values.apply(lambda x:round((x/flights_row)*100,2))
Mis_Vals_Fl.sort_values(by='%_Missing',ascending=False)
```

```python
drop_cols_flights = Mis_Vals_Fl[Mis_Vals_Fl['%_Missing']> 80]
drop_cols_flights.index
```

`CANCELLATION_REASON`, `AIR_SYSTEM_DELAY`, `SECURITY_DELAY`,`AIRLINE_DELAY`, `LATE_AIRCRAFT_DELAY`, `WEATHER_DELAY` Columns had more than 80% of the missing values, hence we could drop these columns

```python
#Dropping columns with more than 80% of missing values
flights.drop(labels = drop_cols_flights.index.tolist(),axis=1, inplace=True)
```

```python
Mis_Vals_Fl[Mis_Vals_Fl['%_Missing']> 0].sort_values(by='%_Missing',ascending=False)
```
