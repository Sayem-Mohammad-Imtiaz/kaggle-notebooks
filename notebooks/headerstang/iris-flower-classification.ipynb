```python
y=data.species
X=data.drop(["species"],axis=1)

# normalization
X=(X-np.min(X))/(np.max(X)-np.min(X))

X_sepal=X.loc[:,["sepal_length","sepal_width"]]
X_petal=X.loc[:,["petal_length","petal_width"]]
```

```python
X_sepaltrain,X_sepaltest,y_sepaltrain,y_sepaltest=train_test_split(X_sepal,y,test_size=0.3,random_state=42)
X_petaltrain,X_petaltest,y_petaltrain,y_petaltest=train_test_split(X_petal,y,test_size=0.3,random_state=42)
```

```python
logregsepal = LogisticRegression().fit(X_sepaltrain,y_sepaltrain)
logregpetal= LogisticRegression().fit(X_petaltrain,y_petaltrain)
print("logistic regression accuracy for sepal:",logregsepal.score(X_sepaltest,y_sepaltest))
print("logistic regression accuracy for petal:",logregpetal.score(X_petaltest,y_petaltest))
```

```python
knnsepal= KNeighborsClassifier().fit(X_sepaltest,y_sepaltest)
knnpetal=KNeighborsClassifier().fit(X_petaltest,y_petaltest)
print("KNN accuracy for sepal:",knnsepal.score(X_sepaltest,y_sepaltest))
print("KNN accuracy for petal:",knnpetal.score(X_petaltest,y_petaltest))
```

```python
treesepal= DecisionTreeClassifier().fit(X_sepaltest,y_sepaltest)
treepetal=DecisionTreeClassifier().fit(X_petaltest,y_petaltest)
print("Decision tree accuracy for sepal:",treesepal.score(X_sepaltest,y_sepaltest))
print("Decision tree accuracy for petal:",treepetal.score(X_petaltest,y_petaltest))
```

```python
rfsepal= RandomForestClassifier().fit(X_sepaltest,y_sepaltest)
rfpetal=RandomForestClassifier().fit(X_petaltest,y_petaltest)
print("Random Forest accuracy for sepal:",rfsepal.score(X_sepaltest,y_sepaltest))
print("Random Forest accuracy for petal:",rfpetal.score(X_petaltest,y_petaltest))
```
