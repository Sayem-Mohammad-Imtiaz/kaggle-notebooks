```python
df=pd.read_csv('/kaggle/input/titanic/train.csv')

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df.shape
```

```python
r=['PassengerId','Name','Ticket','Cabin']
df1=df.drop(r,axis=1).copy()
df1.head()
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
df1["Embarked"]=df1["Embarked"].fillna(df1["Embarked"].mode()[0])

df1['Age']=df1['Age'].fillna(df1['Age'].mean())
```

```python
df1.head()
```

```python
cc=['Sex','Embarked']

dummy=pd.get_dummies(df1[cc])
dummy.head()
```

```python
df1=df1.drop(cc,axis=1)

df1
```

```python
df2=pd.concat([df1,dummy],axis=1)

df2.head()
```

```python
x=df2.iloc[:,1:]

x.head()
```

```python
y=df2.iloc[:,0]

y.head()

y.unique()

y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,stratify=y,test_size=0.25)
```

```python
from sklearn.metrics import log_loss
from sklearn.ensemble import VotingClassifier
```

```python
from sklearn import tree
dtc = tree.DecisionTreeClassifier(random_state=0)
dtc.fit(x_train, y_train)

dtc.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier(n_estimators=30)
clf.fit(x_train,y_train)

clf.score(x_test,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)

nb.score(x_test,y_test)
```

```python
model_1 = tree.DecisionTreeClassifier(random_state=0)
model_2 = RandomForestClassifier(n_estimators=30)
model_3 = GaussianNB()
```

```python
final_model = VotingClassifier(
    estimators=[('dtc', model_1), ('clf', model_2), ('nb', model_3)], voting='hard')
```

```python
final_model.fit(x_train, y_train)
```

```python
pred_final = final_model.predict(x_test)
```

```python
print(log_loss(y_test, pred_final))
```
