Let's estimate the anual revenue of each genre by scaling the number of reviews by price. Assuming that the same ratios of people review the books they purchase across genres, this will show the market share of a genre broken down by year.

But first, we have to eliminate duplicate data - books that were bestsellers in multiple years. We should only count their sales once, so let's delete all but the earliest year of sales. For example, see the title of the dataframe below:

```python
M.sort_values(by = ['Name','Year'])
```

```python
N=M.drop_duplicates(subset = ['Name','Author','Reviews']).copy()
N.describe()
```

It worked!

```python
N["Estimated Revenue (In 100,000s)"]=list(N['Price']*N['Reviews']/100000)
    
```

```python
N.describe()
```

```python
N.head()
```

```python
fig = px.bar(N,x='Year',y='Estimated Revenue (In 100,000s)',color='Genre',hover_data = ['Name','Author','Reviews','User Rating','Estimated Revenue (In 100,000s)'],barmode='group')
fig.show()
```

> 

Visualize the relation between price and reviews:

```python
fig = px.scatter(N,x='Price',y='Reviews',color='Genre',hover_data = ['Name','Author','Reviews','Estimated Revenue (In 100,000s)'])
fig.show()
```
