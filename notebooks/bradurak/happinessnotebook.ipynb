## Verilerin Birleştirilmesi

```python
df_c1 = pd.concat([df_2015,df_2016,df_2017])
df_c2 = pd.concat([df_2018,df_2019])
```

```python
scores_15_16 = df_c1['Happiness Score'].dropna()
scores_17 = df_c1['Happiness.Score'].dropna()
apended = scores_15_16.append(scores_17)
df_c1['Happiness Score'] = apended
```

```python
economy_15_16 = df_c1['Economy (GDP per Capita)'].dropna()
economy_17 = df_c1['Economy..GDP.per.Capita.'].dropna()
apended = economy_15_16.append(economy_17)
df_c1['Economy (GDP per Capita)'] = apended
```

```python
health_15_16 = df_c1['Health (Life Expectancy)'].dropna()
health_17 = df_c1['Health..Life.Expectancy.'].dropna()
apended = health_15_16.append(health_17)
df_c1['Health (Life Expectancy)'] = apended
```

```python
dystopia_15_16 = df_c1['Dystopia Residual'].dropna()
dystopia_17 = df_c1['Dystopia.Residual'].dropna()
apended = dystopia_15_16.append(dystopia_17)
df_c1['Dystopia Residual'] = apended
```

```python
government_15_16 = df_c1['Trust (Government Corruption)'].dropna()
government_17 = df_c1['Trust..Government.Corruption.'].dropna()
apended = government_15_16.append(government_17)
df_c1['Trust (Government Corruption)'] = apended
```

```python
happiness_15_16 = df_c1['Happiness Rank'].dropna()
happiness_17 = df_c1['Happiness.Rank'].dropna()
apended = happiness_15_16.append(happiness_17)
df_c1['Happiness Rank'] = apended
```

```python
df_c1.drop(columns = [ 'Happiness.Rank', 'Happiness.Score', 'Economy..GDP.per.Capita.',
       'Health..Life.Expectancy.', 'Trust..Government.Corruption.',
       'Dystopia.Residual','Lower Confidence Interval','Upper Confidence Interval','Region','Standard Error','Whisker.high','Whisker.low'] , axis =1,inplace = True)
```

```python
df_c2.columns = ['Happiness Rank','Country' , 'Happiness Score','Economy (GDP per Capita)','Family', 'Health (Life Expectancy)',
       'Freedom', 'Generosity', 'Trust (Government Corruption)', 'Year']
```

```python
df = df_c1.append(df_c2)
```

```python
df.tail()
```

```python
df.drop(['Dystopia Residual'] , axis = 1 , inplace = True)
```

```python
df.isnull().any()
```

```python
pip install  ycimpute
```

```python
from ycimpute.imputer import knnimput
```

```python
df['IsTurkey'] = df['Country'] == "Turkey"
```

```python
df.head()
```

```python
df['IsTurkey'] = df['IsTurkey'].astype('int64') 
```

```python
df_numerical = df.select_dtypes(include = ["float64","int64",'int16'])
```

```python
knned_df_numerical = np.array(df_numerical)
knned_df_numerical
```

```python
knned_np = knnimput.KNN(k = 4).complete(knned_df_numerical)
```

```python
knned_df = pd.DataFrame(knned_np , columns = df_numerical.columns)
```

```python
knned_df.isnull().any()
```

```python
knned_df.info()
```

```python
df_turkey = knned_df[knned_df['IsTurkey'] == 1]
```

```python
df_turkey
```

```python
df_turkey.drop(['IsTurkey'] ,axis = 1, inplace = True )
df_turkey
```

```python
df_turkey.groupby("Year").sum()["Happiness Score"].plot(color = 'green')
```

```python
sns.barplot(x="Year", y="Happiness Score" , data = df_turkey)
```

```python
df_turkey_score = df_turkey.groupby("Happiness Score").sum()
```

```python
df_turkey_score.reset_index(inplace = True , drop = True)
```
