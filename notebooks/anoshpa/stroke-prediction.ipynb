# Data Preprocessing

```python
# Round off Age
data['age'] = data['age'].apply(lambda x : round(x))
```

```python
#lets see the missing values in the dataset
data.isnull().sum()
```

```python
# so omly `bmi`(body mass index) column have missing values
data["bmi"]
```

```python
# let`s replace missing vales in bmi with mean
bmi_mean=data["bmi"].mean()
data["bmi"]=data["bmi"].fillna(bmi_mean)
```

```python
data["bmi"].isnull().sum() #so no null value now
```

```python
# let`s look at unique values in each column
data.nunique()
```
