## Matplot (Grafiklerin Oluşturulması)

```python
data.age.plot(kind='line', color='r', label='Age', linewidth=1, alpha=1, grid= True, linestyle=':')
plt.legend(loc='lower right')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Line Plot')
plt.show()
```

Yukardaki grafik ile kalp krizinin hangi yaşlarda olduğu analiz edilmiştir. 
Ağırlıklı olarak 45-60 yaş arasında seyr ettiğini grafikler yardımıyla görebiliriz.

```python
data.plot(kind='scatter',x='age', y='trestbps', alpha=1, color='g')
plt.legend(loc='upper right')
plt.xlabel('age')
plt.ylabel('trestbps')
plt.title('Scatter Plot')
plt.show()
```

## WHILE and FOR LOOPS

```python
i = 55
while i != 62 :
    print('i is: ',i)
    i +=1 
print(i,' is equal to 62')
```

```python
lis = [59,60,61,62]
for i in lis:
    print('i is: ',i)
print('')

for index, value in enumerate(lis):
    print(index," : ",value)
print('')   

dictionary = {'age':'35','trestbps':'150'}
for key,value in dictionary.items():
    print(key," : ",value)
print('')

for index,value in data[['age']][0:1].iterrows():
    print(index," : ",value)
```
