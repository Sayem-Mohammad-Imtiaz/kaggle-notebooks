# Heatmap of Categories Group by Years

```python
heatmapData = (df
 .groupby('Year')
 .category_name_1
 .value_counts()
 .unstack()
 .fillna(0)
)

labels_x = ['2016','2017','2018']
labels_y = ['Appliances','Beauty & Grooming','Books','Computing','Entertainment','Health & Sports','Home & Living','Kids & Baby',
            'Mens Fashion','Mobiles & Tablets','Others','School & Education','Gourmet & Dried Fruit','Superstore','Womens Fashion','Not Available']

plt.figure(figsize=(25,10))

sns.heatmap(heatmapData.T, 
            cmap = 'YlGnBu', 
            annot = True, 
            fmt = '.0f', 
            center = 30000,
            linecolor = 'black',
            xticklabels = labels_x,
            yticklabels = labels_y,
           annot_kws ={'fontsize': 16,}
           )
```

# Mobiles & Tablets are Best Selling category in Ecommerce.

**Top 3 Categroies in 2016:**
* Men's Fashion
* Mobiles & Tablets
* Gourmet & Dried Fruit

**Top 3 Categroies in 2017:**
* Mobiles & Tablets
* Men's Fashion
* Women's Fashion

**Top 3 Categroies in 2018:**
* Mobiles & Tablets
* Others
* Men's Fashion

#  Most Popular Payment Methods

```python
heatmapPayment = (df
 .groupby('payment_method')
 .category_name_1
 .value_counts()
 .unstack()
 .fillna(0)
)

labels_y = ['Appliances','Beauty & Grooming','Books','Computing','Entertainment','Health & Sports','Home & Living','Kids & Baby',
            'Mens Fashion','Mobiles & Tablets','Others','School & Education','Gourmet & Dried Fruit','Superstore','Womens Fashion','Not Available']

plt.figure(figsize=(25,10))

sns.heatmap(heatmapPayment.T, 
            cmap = 'Blues', 
            annot = True, 
            fmt = '.0f', 
            yticklabels = labels_y,
           annot_kws ={'fontsize': 16,}
           )
```

This heatmap shows us that **COD** is the **most popular payment method**.

# Best Selling Categories as per Revenue

```python
plt.figure(figsize=(30,15))

sns.set_style("whitegrid")
sns.barplot(x='price', y='category_name_1', data = df, ci=None, palette='YlGnBu', hue='Year')
```

**Top 3 Categories as per revenues:**

* Mobiles & Tablets
* Entertainment
* Computing

**THE NET AND GROSS SELL OF MOIBLES & TABLETS CATEGORY**

```python
MT_Gross = df[df['category_name_1'].isin(['Mobiles & Tablets'])]
MT_Gross.price.sum()
```

> **Mobiles & Tablets** one of the famous category with the total gross revenue of **PKR 2,030,105,565**

```python
MT_Net = df[(df.status == 'complete') | (df.status == 'paid') | (df.status == 'received')]
MT_Net.price.sum()
```

> **Mobiles & Tablets** one of the famous category with the total net revenue of **PKR 1,231,196,784**
