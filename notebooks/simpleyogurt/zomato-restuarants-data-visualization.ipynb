```python
import seaborn as sns
import matplotlib.pyplot as plt
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

np.warnings.filterwarnings('ignore')
```

```python
sns.countplot(dff['online_order'])
plt.title("Restaurants have online order?")
```

```python
labels=["Yes" , "No"]
values=dff["online_order"].value_counts().values
fig1, ax1 = plt.subplots()
ax1.pie(values, labels=labels,autopct='%1.1f%%',shadow=True, startangle=90)
plt.title('Online order Accepted?')
plt.show()
```

```python
sns.countplot(dff['book_table'])
plt.title("Restaurants  have table booking?")
```

```python
labels=["No" , "Yes"]
values=dff["book_table"].value_counts().values
fig1, ax1 = plt.subplots()
ax1.pie(values, labels=labels,autopct='%1.1f%%',shadow=True, startangle=90)
plt.title('Restaurants have have table booking?')
plt.show()
```
