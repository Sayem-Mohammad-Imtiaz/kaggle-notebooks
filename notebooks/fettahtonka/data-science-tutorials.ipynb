# Python Data Science Toolbox

```python
# lambda function
area_triangle=lambda x,y:(x*y)/2
print("triangle's area :",area_triangle(5,4))

area_circle=lambda pi,r:pi*(r*r)
print("circle's area :",area_circle(3.14,10))
```

```python
# anonymous function
radius=[2.5,5,7.5,10]
areas=map(lambda r:3.14*(r*r),radius)
print(list(areas))
```

```python
# zip method
list1=[1,2,3,4,5,6,7]
list2=[ "Africa", "Antarctica", "Asia", "Australia", "Europe", "North America", "South America"]
zipped=zip(list1,list2)
listed=list(zipped)
listed
```

```python
# un-zip
un_zip=zip(*listed)
un_list1,un_list2=list(un_zip)
print(un_list1)
print(un_list2)
```

```python
# list comprehension
numbers1=[1,2,3,4,5]
numbers2=[i**2 for i in numbers1]
numbers2
```

```python
# conditional list comprehension
numbers3=[20,40,60,80,100]
numbers4=[i-i if i<50 else i+i for i in numbers3]
numbers4
```
