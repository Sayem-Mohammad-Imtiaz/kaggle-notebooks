**3. Obtén una lista con los atributos cuantitativos y otra lista con los atributos cualitativos.**

```python
tipos=df.dtypes
cuantitativos=list()
cualitativos=list()
tipos
for i in range(len(tipos)):
    if tipos[i] == "object":
        cualitativos.append(tipos.index[i])
    else:
        cuantitativos.append(tipos.index[i])
print(cuantitativos)
print(cualitativos)
```
