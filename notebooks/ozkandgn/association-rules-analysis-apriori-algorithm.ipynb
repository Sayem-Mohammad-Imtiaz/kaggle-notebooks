## Association Rules Analysis - Apriori Algorithm

### Explanation
- In this project, relations were analyzed between products.
- It was analyzed which other products were purchased with the purchased product.

```python
import pandas as pd
```

```python
### For apriori algorithm

!pip install mlxtend
```

```python
### Groceries dataset

data = pd.read_csv("../input/groceries/groceries.csv",error_bad_lines=False,header=None);
```

```python
### Count of the values

data.apply(data.value_counts).sum(axis=1).sort_values(ascending=False)
```

```python
### Pandas Dataframe to list

### We need product lists for apriori algorithm

records = []
for i in range((len(data))):
    records.append([str(data.values[i,j]) for j in range(0, data.shape[1])])
records[:10]
```

```python
### True - False (Dummy) array

from mlxtend.preprocessing.transactionencoder import TransactionEncoder

te = TransactionEncoder()
te_data = te.fit(records).transform(records)
df = pd.DataFrame(te_data, columns = te.columns_)
df = df.drop(("nan"),axis=1)
df
```

### Support value
- Ratio of product

Support(X, Y) = Freq(X,Y)/N

```python
### Apriori algorithm
### Coexistence rate of products


from mlxtend.frequent_patterns import apriori

df1 = apriori(df, min_support=0.003, use_colnames = True, verbose=1)
df1
```

### Antecedents
- First product to purchase

### Consequents
- Second product that can be purchase with first product

### Confidence
- Ratio of purchasers of the product to purchase the other product

### Lift 
- Ratio of purchase increase second product owing to fist product purchasing



```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(df1, metric = "confidence", min_threshold = 0.2)
rules
```

```python
rules[rules["lift"]>2]
```
