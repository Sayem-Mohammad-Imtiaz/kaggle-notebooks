In this notebook, I'm going to use Linear regression method to predict price of unit area.

first of all I want to give you some information about this dataset I'm going to use, this dataset includes data on features of 414 different houses like 'transaction date', 'house age','distance to the nearest MRT station','price of unit area' etc. 

As I said before, I want to predict price of unit area, this prediction is going to be base on three independent variavble, 'house age',  'distance to the nearest MRT station', 'number of convenience stores'.

Let's get started.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Reading the dataset

```python
df = pd.read_csv('../input/real-estate-price-prediction/Real estate.csv')
```

# Data overview

```python
df.head()
```

```python
df.columns
```

# preparing dataset to modeling
Creating dataframe that include only Required features

```python
re = df.drop(['No', 'X1 transaction date', 'X5 latitude', 'X6 longitude'], axis = 1)
```

```python
re.head()
```

```python
re.corr()
```

```python
sns.pairplot(re)
```

# determining features and label

X refers to features and y refers to label

```python
X = re.drop(['Y house price of unit area'], axis  = 1)
y = re['Y house price of unit area']
```

# splitting dataset into two different partition for training/testing data


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=101)
```

```python
y_test
```

```python
X_test
```

# fitting a model to training data

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
pd.DataFrame(model.coef_, X.columns, columns=['coeficients'])
```

As it is computed above, if house age increase one unit, price of unit area will decrese 0.27 unit.

if distance to the nearest MRT station increase one unit, price of unit area almost will be same as before.

if number of convenience stores increase one unit, price of unit area will increase 1.41 unit.


# Predicting price of unit area

```python
y_pred = model.predict(X_test)
```

# Evaluating the model

```python
from sklearn import metrics
MAE = metrics.mean_absolute_error(y_test, y_pred) # MAE is short for Mean-Absolute-Error
MAE  
```

```python
MSE = metrics.mean_squared_error(y_test, y_pred)  # MSE is short for Mean-Square-Error
MSE
```

```python
RMSE = np.sqrt(MSE)  # RMSE is short for Root-Mean-Square-Error
RMSE
```

```python
re['Y house price of unit area'].mean()
```

metrics evaluating in compared with mean of prices of unit area show this prediction has too much Error so this model is not good way to predict price.

```python
test_residuals = y_test - y_pred
```

```python
sns.displot(data = test_residuals, bins = 30, kde = True)
```

```python
sns.scatterplot(x = y_test, y = test_residuals)
plt.axhline(y = 0, c = 'r', ls = '--')
```

As you see in the above scatter plot, density of points in the center of plot is higher than other parts. 

it shows that **linear regression** is not appropriate way to predict Price of unit area based on 'house age', 'distance to the nearest MRT station' and 'number of convenience stores'.

Note: If you take a quick look at pairplot that plotted in the middle of notebook, you immediately find that there is no linear relationship between none of independent variables and dependent variable. due to that, linear regression is not appropriate way to predict house price of unit area.
