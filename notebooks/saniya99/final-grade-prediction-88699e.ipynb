```python
import numpy as np
import pandas as pd 
import seaborn as sns
# import os
# print(os.listdir("../input"))
from sklearn.utils import shuffle
```

```python
data1 = pd.read_csv("../input/student-mat.csv",sep=",")
data2 = pd.read_csv("../input/student-por.csv",sep=",")
data = [data1,data2]
data=pd.concat(data)
data=shuffle(data)
data.describe(include="all")
```

```python
data=data.drop_duplicates(["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"])
data.describe(include="all")
```

Removing duplicates leads to 662 tuples, which implies there were 382 duplicate tuples in the two datasets

```python
# print(data.isnull().count())
data.replace([np.inf, -np.inf], np.nan, inplace=True) 
data = data.dropna()
data.dropna(inplace=True) 
data.describe(include="all")
```

So from here we can conclude, that there were no tuples with NULL or infinite values.

```python
sns.catplot(x="school", hue = "sex" , data=data , kind="count",height=6, aspect=.7)
print("percentage total female : ",(data["sex"] == 'F').value_counts(normalize = True)[1]*100)
print("percentage total male : ",(data["sex"] == 'M').value_counts(normalize = True)[1]*100)
```

It is clear that number of female students are more than number of male students in both the  schools, with 'GP' - Gabriel Pereira school having more number of students overall.

```python
#drop some features that have very less correlation values 
datagrade = data.drop(["traveltime","famrel","freetime","goout","health","absences"], axis=1)
data_alc = data.drop(["traveltime","famrel", "Fedu", "Medu","goout","health","absences"], axis=1)
```

```python
#to suppress the warnings
import sys
import warnings

if not sys.warnoptions:
    warnings.simplefilter("ignore")
```

```python
#data cleaning(binary and onehot encodings)
datagrade.school[datagrade.school == 'GP'] = 1
datagrade.school[datagrade.school == 'MS'] = 0
datagrade.sex[datagrade.sex == 'M'] = 1
datagrade.sex[datagrade.sex == 'F'] = 0
datagrade.address[datagrade.address == 'U'] = 1
datagrade.address[datagrade.address == 'R'] = 0
datagrade.famsize[datagrade.famsize == 'GT3'] = 1
datagrade.famsize[datagrade.famsize == 'LE3'] = 0
datagrade.Pstatus[datagrade.Pstatus == 'T'] = 1
datagrade.Pstatus[datagrade.Pstatus == 'A'] = 0
#for Medu & Fedu its better in numeric as higher education has higher difference with primary than secondary
#one hot encoding of necessary features
cols_to_transform = [ 'Mjob','Fjob','reason','guardian' ]
datagrade=pd.get_dummies(datagrade,columns=cols_to_transform) 
#failures, traveltime, studytime does not require onehot  
```

```python
datagrade.schoolsup[datagrade.schoolsup == 'yes'] = 1
datagrade.schoolsup[datagrade.schoolsup == 'no'] = 0
datagrade.famsup[datagrade.famsup == 'yes'] = 1
datagrade.famsup[datagrade.famsup == 'no'] = 0
datagrade.paid[datagrade.paid == 'yes'] = 1
datagrade.paid[datagrade.paid == 'no'] = 0
datagrade.activities[datagrade.activities == 'yes'] = 1
datagrade.activities[datagrade.activities == 'no'] = 0
datagrade.nursery[datagrade.nursery == 'yes'] = 1
datagrade.nursery[datagrade.nursery == 'no'] = 0
datagrade.higher[datagrade.higher == 'yes'] = 1
datagrade.higher[datagrade.higher == 'no'] = 0
datagrade.internet[datagrade.internet == 'yes'] = 1
datagrade.internet[datagrade.internet == 'no'] = 0
datagrade.romantic[datagrade.romantic == 'yes'] = 1
datagrade.romantic[datagrade.romantic == 'no'] = 0

#remaining doesn't require onehot as distances are not same. one hot is used only when they are equidistant
```

```python
#data cleaning for alcohol consumption prediction objective
data_alc.school[data_alc.school == 'GP'] = 1
data_alc.school[data_alc.school == 'MS'] = 0
data_alc.sex[data_alc.sex == 'M'] = 1
data_alc.sex[data_alc.sex == 'F'] = 0
data_alc.address[data_alc.address == 'U'] = 1
data_alc.address[data_alc.address == 'R'] = 0
data_alc.famsize[data_alc.famsize == 'GT3'] = 1
data_alc.famsize[data_alc.famsize == 'LE3'] = 0
data_alc.Pstatus[data_alc.Pstatus == 'T'] = 1
data_alc.Pstatus[data_alc.Pstatus == 'A'] = 0


cols_to_transform = [ 'Mjob','Fjob','reason','guardian' ]
data_alc=pd.get_dummies(data_alc,columns=cols_to_transform) 
```

```python
data_alc.schoolsup[data_alc.schoolsup == 'yes'] = 1
data_alc.schoolsup[data_alc.schoolsup == 'no'] = 0
data_alc.famsup[data_alc.famsup == 'yes'] = 1
data_alc.famsup[data_alc.famsup == 'no'] = 0
data_alc.paid[data_alc.paid == 'yes'] = 1
data_alc.paid[data_alc.paid == 'no'] = 0
data_alc.activities[data_alc.activities == 'yes'] = 1
data_alc.activities[data_alc.activities == 'no'] = 0
data_alc.nursery[data_alc.nursery == 'yes'] = 1
data_alc.nursery[data_alc.nursery == 'no'] = 0
data_alc.higher[data_alc.higher == 'yes'] = 1
data_alc.higher[data_alc.higher == 'no'] = 0
data_alc.internet[data_alc.internet == 'yes'] = 1
data_alc.internet[data_alc.internet == 'no'] = 0
data_alc.romantic[data_alc.romantic == 'yes'] = 1
data_alc.romantic[data_alc.romantic == 'no'] = 0

#remaining doesn't require onehot as distances are not same. one hot is used only when they are equidistant
```

```python
def convertToGrades(x):
    if x>=18 and x<=20:
        return 5  #A grade
    elif x>15:
        return 4  #B grade
    elif x>10:
        return 3  #C grade
    elif x>5:
        return 2  #D grade
    else:
        return 1  #E grade
datagrade.G3 = datagrade.G3.map(convertToGrades)

data.head(10)
```

```python
#Final grades
y_grade =  datagrade[[ 'G3']].mean(axis=1)
data_grade = datagrade.drop(["G3"], axis=1)
FEATURE_GRADE = ['Index', 'School', 'Gender', 'Age', 'Address', 'Family Size', 'Parents status', 'Mother Education', 'Father Education', 'Study Time', 'Failures', 'SchoolSup', 'FamSup', 'Paid', 'Activites', 'Nursery', 'Higher Studies', 'Internet', 'Romantic Relationships', 'Weekday Alcohol Consumption', 'Weekend Alcohol Consumption', '1st Period Results', '2nd Period Results', 'Does Mother work at home', 'Does Mother work in health sector', 'Does Mother work somewhere else', 'Does Mother work in administrative services', 'Is mother teacher','Does Father work at home', 'Does Father work in health sector', 'Does Father work somewhere else', 'Does Father work in administrative services', 'Is father teacher' 'Reason Course', 'Reason Home', 'Reason Other', 'Reason Reputation', 'Guardian Father', 'Guardian Mother', 'Guardian Other']
data_grade.head()
```

```python
#Alchohol Consumption
y_alc_weekday =  data_alc[[ 'Dalc']].mean(axis=1)
y_alc_weekend =  data_alc[[ 'Walc']].mean(axis=1)
data_alc = data_alc.drop(["Dalc", "Walc"], axis=1)
FEATURE_ALC = ['Index', 'School', 'Gender', 'Age', 'Address', 'Family Size', 'Parents status', 'Study Time', 'Failures', 'SchoolSup', 'FamSup', 'Paid', 'Activites', 'Nursery', 'Higher Studies', 'Internet', 'Romantic Relationships', 'Freetime', '1st Period Results', '2nd Period Results', 'Final results', 'Does Mother work at home', 'Does Mother work in health sector', 'Does Mother work somewhere else', 'Does Mother work in administrative services', 'Is mother teacher','Does Father work at home', 'Does Father work in health sector', 'Does Father work somewhere else', 'Does Father work in administrative services', 'Is father teacher' 'Reason Course', 'Reason Home', 'Reason Other', 'Reason Reputation', 'Guardian Father', 'Guardian Mother', 'Guardian Other']
data_alc.head()
```

```python
#spliting data to train and test by 80% and 20% respectievely 
from sklearn.model_selection import train_test_split
X_grade_train, X_grade_test, y_grade_train, y_grade_test = train_test_split(data_grade,y_grade, test_size=0.2)
```

```python
#spliting data to train and test by 80% and 20% respectievely 
X_alc_weekday_train, X_alc_weekday_test, y_alc_weekday_train, y_alc_weekday_test = train_test_split(data_alc,y_alc_weekday, test_size=0.2)
X_alc_weekend_train, X_alc_weekend_test, y_alc_weekend_train, y_alc_weekend_test = train_test_split(data_alc,y_alc_weekend, test_size=0.2)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, classification_report, confusion_matrix
from sklearn.tree import export_graphviz
import matplotlib.pyplot as plt
import cv2


dtree_model = DecisionTreeClassifier(max_depth = 2).fit(X_grade_train, y_grade_train) 
dtree_predictions = dtree_model.predict(X_grade_test) 
export_graphviz(dtree_model, 'tree.dot', feature_names = FEATURE_GRADE)  
! dot -Tpng tree.dot -o tree.png
%matplotlib inline
img = cv2.imread('tree.png')
plt.figure(figsize = (20, 20))
plt.imshow(img)
acc = f1_score(y_grade_test, dtree_predictions, average="macro")
print("F1 score for Predicting final grade using decision tree of max 2 depth: " + str(acc))






```
