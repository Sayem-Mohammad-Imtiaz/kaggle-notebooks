# Cleaning the "Book_Name" column

```python
Final["Book_Name"] = Final["Book_Name"].str.split("/")
Final = Final.explode("Book_Name", ignore_index=True)
Final["Book_Name"] = Final["Book_Name"].str.split("?")
Final = Final.explode("Book_Name", ignore_index=True)
Final["Book_Name"] = Final["Book_Name"].str.split("- ")
Final = Final.explode("Book_Name", ignore_index=True)
Final= Final.replace(to_replace=r"^\s*$", value=np.NAN, regex=True)
Final =Final.replace(to_replace=r"^[!(-. \d]*$", value=np.NAN, regex=True)
Final= Final.replace(to_replace=r"^Release Date$", value=np.NAN, regex=True)
Final = Final.dropna(subset=["Book_Name"])
#print(Final.isna().sum())
```

# Analyze the Top_10 Sold Books

```python
#Analyze data
Top_10_books = Final.groupby("Book_Name")["Order_NO"].agg(["count"]).reset_index().sort_values(by="count",ascending=False).head(10)
#print(Top_10_books)
```

```python
#Visuallize the Top ten Books
sns.set_style("darkgrid")
sns.set_context("paper")
g=sns.catplot(x="Book_Name",y="count",data=Top_10_books,kind="bar")
g.fig.suptitle("Top 10 Books ")
plt.xticks(rotation=90)
plt.show()
```

# Analyze the Top_15 cities in which books are sold

# Order of Books by Year

```python
# Check the null vales in the column
Final["Order_Date"].isnull().sum()
```

```python
# Analyze the order of Books BY Year

Final["Order_Date"]=pd.to_datetime(Final["Order_Date"])
Final["Year"]=Final["Order_Date"].dt.year
Per_Year=Final.groupby("Year")["Order_NO"].agg(['count']).reset_index().sort_values(by='count',ascending=False)
print(Per_Year)
```

```python
# Visualize the Graph

sns.set_style("darkgrid")
sns.set_context("paper", rc={"lines.linewidth": 2})
g=sns.barplot(x="Year",y="count",data=Per_Year,)
g.set_title("Order By Years")
plt.xticks(rotation=90)
plt.show()
```

# Order of Books by Month

```python
# Analyze the order of books by Months
Final["Month"] = Final["Order_Date"].dt.month_name()
month=["January","February ","March","April","May","June","July","August","September","October","November","December"]
Per_Month_stat=Final.groupby("Month")["Order_NO"].agg(['count']).reset_index().sort_values(by="count",ascending=False)
print(Per_Month_stat)
```

```python
#Visualize Bar plot show the no of the order of books by Months

cat_order=["January","February","March","April","May","June",
          "July","August","September","November","December"]
sns.set_style("darkgrid")
sns.set_context("paper", rc={"lines.linewidth": 2})
g=sns.barplot(x="Month",y="count",data=Per_Month_stat,order=cat_order)
g.set_title("Orders By Months ")
plt.xticks(rotation=90)
plt.show()
```

```python
#Visualize the relational graph of the order of books by Months
cat_order=["January","February","March","April","May","June",
          "July","August","September","November","December"]
sns.set_style("whitegrid")
sns.set_context("paper", rc={"lines.linewidth": 2})
g=sns.catplot(x="Month",y="count",kind='point',data=Per_Month_stat,order=cat_order,color="r")
g.fig.suptitle("Orders BY Months ")
plt.xticks(rotation=90)
plt.show()
```

# Relation of order of books by months and year
