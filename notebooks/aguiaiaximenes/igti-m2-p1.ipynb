**Verificando se existem outliers (anomalias) nos dados. Para isso, será utilizado o boxplot**

```python
boxplot = customers.boxplot(column=['Age', 'Annual Income (k$)', 'Spending Score (1-100)']) 
#constroi o boxplot para as colunas desejadas
```

* Verde = mediana
* Quartis = da primeira linha preta à primeira linha azul = 1º quartil; da primeira linha azul até a verde = 2º quartil; da linha verde até a segunda linha azul = 3º quartil; da segunda linha azul até a última linha preta = 4º quartil
* Dentro da caixa azul, portanto, temos 50% dos nossos dados
* Bolinha --> indicação de um outlier

**Verificando se existem outliers (anomalias) nos dados. Para isso será utilizado o Z-Score**

```python
#z-score
from scipy import stats
z = np.abs(stats.zscore(customers['Annual Income (k$)'].values))
thereshold = 2
result=np.where(z > thereshold)

df_salario_outlier=customers.iloc[result[0]]
#print(z)
```
