### 1 - Parental level of education has correlation with the score?

```python
df.groupby('parental level of education')[['math score', 'reading score','writing score']].mean().plot(kind='bar')
plt.margins(x=0, y=-0.0)   
```

In this graphic we see that high school has the minor score mean and the master has the best score.

### 2. Race/ethnicity has correlation with the score?

```python
df.groupby('race/ethnicity')[['math score', 'reading score','writing score']].mean().plot(kind='bar')
plt.margins(x=0, y=-0.0)   
```

### 3. Is it has test preparation course did doing difference?

```python
df.groupby('test preparation course')[['math score', 'reading score','writing score']].mean().plot(kind='bar')
plt.margins(x=0, y=-0.0)   
```

Students with complete test preparation course has high score in average. 

### 4. Male or female, who have the best score?

I'm looking for differences.

```python
dfvl = df['gender'].value_counts().plot(kind='bar')
plt.grid(True)
```

```python
df.groupby('gender')[['math score', 'reading score','writing score']].mean().plot(kind='bar')
plt.margins(x=0, y=-0.0)   
plt.grid(True)
```

```python
sms = df[df['gender']=='male'].groupby('gender')['math score'].sum()
sfs =  df[df['gender']=='female'].groupby('gender')['math score'].sum()

print(sms[0] * 100 / sfs[0] / 100)
```

We see, male has 1,005005764% more scores in math. But in the others areas, female has the high scores.
