```python
data = df.drop(['RESALE','READY_TO_MOVE','LONGITUDE','LATITUDE','ADDRESS','POSTED_BY','UNDER_CONSTRUCTION'],axis=1)
data.head()
```

```python
# label encoding

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['BHK_OR_RK']=le.fit_transform(data['BHK_OR_RK'])
```

```python
data.head()
```

```python
# skewness in target

data.skew(axis = 0, skipna = True) 
```

```python
data['TARGET(PRICE_IN_LACS)'] = np.log(data['TARGET(PRICE_IN_LACS)'])
```

```python
data.skew(axis =0)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X = data.drop('TARGET(PRICE_IN_LACS)',axis=1)
Y = data['TARGET(PRICE_IN_LACS)']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)

linear_reg = LinearRegression()

model=linear_reg.fit(X_train,Y_train)
pred = model.predict(X_test)
```

```python
# Checkiing accuracy


from sklearn.metrics import r2_score

r2_score(Y_test,pred)
```

```python
from sklearn.metrics import mean_squared_error
from math import *

sqrt(mean_squared_error(pred,Y_test))
```
