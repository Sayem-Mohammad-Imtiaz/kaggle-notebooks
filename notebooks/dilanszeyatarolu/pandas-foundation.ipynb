VISUAL EXPLORATORY DATA ANALYSIS


```python
 # Plot

data1=data.loc[:,["Confirmed","Deaths","Recovered"]]
data1.plot()
```

```python
# subplot
data1.plot(subplots=True)
plt.show()
```

```python
# Scatter plot

data1.plot(kind="scatter",x="Deaths",y="Confirmed")
plt.show()
```

```python
#Histogram Plot

data2=data.loc[:,["Recovered","Deaths","SNo"]]
data2.plot(kind="hist",y="Deaths",range=(0,1000),bins=10)
plt.show()
```

```python
data2.plot(kind="hist",y='Deaths',cumulative=True,range=(0,100))
plt.show()
```
