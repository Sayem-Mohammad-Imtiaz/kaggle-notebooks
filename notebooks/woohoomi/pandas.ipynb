```python
#统计电影分类
df = pd.read_csv('/kaggle/input/imdb-data/IMDB-Movie-Data.csv')
templist = df['Genre'].str.split(',').tolist()
#嵌套列表展开 @todo
genre_list = list(set(i for j in templist for i in j))
genre_list
```

```python
#生成一个空的矩阵
zero_df = pd.DataFrame(np.zeros((df.shape[0],len(genre_list))),columns=genre_list)
zero_df
```

```python

#每个电影出现分类的位置赋值
#获得df的行数 df.shape[0]
for i in range(df.shape[0]):
    zero_df.loc[i,templist[i]]=1
zero_df
```

```python
#各种电影的统计总量
genre_count = zero_df.sum()
genre_count
```

```python
genre_count = genre_count.sort_values()
genre_count
```
