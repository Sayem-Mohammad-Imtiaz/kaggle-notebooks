```python
codes = [countries[country] if country != 'I prefer not to say' else None for country in data['Country']]
```

```python
data['Code']=codes
```

```python

gdata = [ dict(
        type = 'choropleth',
        locations = data['Code'].value_counts().index,
        z = data['Code'].value_counts(),
        text = data['Code'].value_counts().index,
        colorscale = [[0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
            [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            tickprefix = '#',
            title = '# Devs'),
      ) ]

layout = dict(
    title = 'Developers Distribution',
    geo = dict(
            projection = dict(
                type = 'Mercator'
            ),
            showframe=False
            )
)




figure = dict( data=gdata, layout=layout )
iplot(figure)
```

###Programming Language

```python
languages = [lang for sublist in [str(langs).replace(" ", "").split(";") for langs in data['HaveWorkedLanguage']] for lang in sublist]
languages = pd.Series(languages)
```

```python
languages.value_counts()[0:10]
```

```python
Java = [1 if 'Java' in str(languages) else 0 for languages in data['HaveWorkedLanguage'] ]
Python = [1 if 'Python' in str(languages) else 0 for languages in data['HaveWorkedLanguage'] ]
data['Java']=Java
data['Python'] = Python
```

```python
gdata = [ dict(
        type = 'choropleth',
        locations = data[['Code','Java']].groupby('Code').mean().index,
        z = data[['Code','Java']].groupby('Code').mean()['Java'],
        text = data[['Code','Java']].groupby('Code').mean().index,
        colorscale = [[0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
            [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            tickprefix = '#',
            title = 'Java'),
      ) ]

layout = dict(
    title = 'Java % in Country',
    geo = dict(
            projection = dict(
                type = 'Mercator'
            ),
            showframe=False
            )
)



figure = dict( data=gdata, layout=layout )
iplot(figure)
```
