<a id = '#LoadData'>1. a. Load Data</a><br>

```python
#read the data
```

```python
input_data = pd.read_csv('../input/bank.csv')
```

```python
print(input_data.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
input_data['deposit_y_n'] = input_data['deposit'].map({'yes':1, 'no':0})
```

```python
#we define the input and output variables
output_var = 'deposit_y_n'
```

```python
#lets use only the numeric cols for input vars
input_vars = [col for col in input_data.columns if input_data[col].dtype != 'object']
```

```python
#the list of input vars also has our output var; lets remove it
input_vars.remove(output_var)
```

```python
#before we build any models, we need to have a test set to check the goodness of our models
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(input_data[input_vars], input_data[output_var], test_size = 0.2)
```

```python
print(X_train.shape)
print(X_val.shape)
print(y_train.shape)
print(y_val.shape)
```

# Build a simple Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt1 = DecisionTreeClassifier()
```

```python
dt1.fit(X=X_train[input_vars], y=y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_pred=dt1.predict(X_train[input_vars]), y_true=y_train))
```

```python
print(accuracy_score(y_pred=dt1.predict(X_val[input_vars]), y_true=y_val))
```

<a id = '#RandomForest'>3.b. Random Forest</a><br>

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest1 = RandomForestClassifier()
```

```python
random_forest1.fit(X=X_train[input_vars], y=y_train)
```

```python
#lets get the accuracy
print(accuracy_score(y_pred=random_forest1.predict(X=X_train[input_vars]), y_true=y_train))
```

```python
print(accuracy_score(y_pred=random_forest1.predict(X=X_val[input_vars]), y_true=y_val))
```

# Our validation set accuracy is 80% which is better than the Decision Tree result of 73%

```python
#wow - we have got an accuracy of 98.5% using the default settings; can we get to 100%?
```

```python
random_forest2 = RandomForestClassifier(n_estimators=100, oob_score=True)
```

```python
random_forest2.fit(X=X_train[input_vars], y=y_train)
```

```python
print(accuracy_score(y_pred=random_forest2.predict(X=X_train[input_vars]), y_true=y_train))
```

```python
print(accuracy_score(y_pred=random_forest2.predict(X=X_val[input_vars]), y_true=y_val))
```
