```python
import pandas as pd


def load_df():
    competitions_df = pd.read_csv('../input/meta-kaggle/Competitions.csv')
    competition_tags_df = pd.read_csv('../input/meta-kaggle/CompetitionTags.csv')
    tags_df = pd.read_csv('../input/meta-kaggle/Tags.csv')
    return competitions_df, competition_tags_df, tags_df


def concat_df(competitions_df, competition_tags_df, tags_df):
    competitions_df = competitions_df.query('CanQualifyTiers==True')[['Id', 'Title', 'DeadlineDate', 'EvaluationAlgorithmName']]
    competitions_df = pd.merge(competitions_df,
                               competition_tags_df,
                               left_on='Id',
                               right_on='CompetitionId',
                               suffixes=('_comps', '_tags'))
    competitions_df = pd.merge(competitions_df[['Title', 'DeadlineDate', 'TagId', 'EvaluationAlgorithmName']],
                               tags_df[['Id', 'Name', 'FullPath']],
                               left_on='TagId',
                               right_on='Id')
    competitions_df['DeadlineDate'] = pd.to_datetime(competitions_df['DeadlineDate'])
    competitions_df = competitions_df.sort_values('DeadlineDate')
    competitions_df = competitions_df[competitions_df['FullPath'].str.contains('data type')]
    competitions_df['year'] = competitions_df['DeadlineDate'].dt.year.values
    return competitions_df.reset_index(drop=True)
```

```python
competitions_df, competition_tags_df, tags_df = load_df()
df = concat_df(competitions_df, competition_tags_df, tags_df)
```
