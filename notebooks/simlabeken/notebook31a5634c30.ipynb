NESTEDED FUNCTION
function inside function

```python
def square():
    """return to square function"""
    def add():
        """return to add function"""
        x = 2
        y = 3
        z = x+y
        return z
    return add()**2
print(square())
```

DEFAULT and FLEXEBLE ARGUMENTS

```python
def f(a, b = 1, c = 2):
    y = a+b+c
    return y
print(f(5))
#what if we want change default arguments
print(f(5,4,3))
```

LAMBDA FUNCTION
faster way of wrting function

```python
square = lambda x : x**2
print(square(4))
tot = lambda x, y, z: x+y+z
print(tot(1,2,3))
```

ANONYMOUS FUNCTION
like a lamda function bur it can take more than one arguments

```python
number_list = [1,2,3]
y = map(lambda x:x**2, number_list)
print(list(y))
```

ITERATORS

```python
#zip example
list1 = [1,2,3,4]
list2 = [5,6,7,8]
z = zip(list1, list2)
print(z)
z_list = list(z)
print(z_list)
```

```python
#unzip
un_zip = zip(*z_list)
un_list1, un_list2 = list(un_zip)
print(un_list1)
print(un_list2)
print(type(un_list1))
print(type(un_list2))
```
