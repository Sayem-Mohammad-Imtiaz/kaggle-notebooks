```python
def age_avg(age):
    if age>=70 :
        age_cat='old'
    elif age>50 and age<70:
        age_cat='mid_old'
    elif age>30 and age<50:
        age_cat='elder'
    else:
        age_cat='young'
    return age_cat
```

```python
df_upd['age_category']=df_upd['age'].apply(age_avg)
```

```python
sns.countplot(df_upd['age_category'],hue=df_upd['card'])
```

```python
df_50=df_upd[(df_upd.age<=50) & (df_upd.age>18)]
```

```python
model_data=pd.get_dummies(df_50_dep_rep,columns={'card','age_category'})
```
