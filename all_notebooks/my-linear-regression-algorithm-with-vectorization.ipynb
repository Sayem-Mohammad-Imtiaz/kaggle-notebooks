```python
import numpy as np # for numerical operations
import pandas as pd # for handling input data
import matplotlib.pyplot as plt # for data visualization 
train = pd.read_csv('../input/bmi-calculation/BMI_Dataset_train.csv') # uses pandas library to open a .csv file
test = pd.read_csv('../input/bmi-calculation/BMI_Dataset_test.csv')
```

```python
test.head()
```

```python
x= train.drop(['BMI','ID'],axis=1)
label=train['BMI']
```

```python
x=np.array(x)
y=np.array(label)
```

```python
w=np.ones((1,x.shape[1]))
print(w)
```

```python
w.shape
```

```python
def hypo(w,x,b):#prediction
    pred=np.dot(w,x.T)+b
    return pred
```

```python
def error(pred,y,m):#cost #error
    er=(1/(2*m))*np.sum(np.square(pred-y.T))
    return er
```

```python
def gradient(pred,y,m,alpha,w,b):#gradient decent
    dw=(1/m)*np.dot((pred-y.T),x)
    db=(1/m)*np.sum(pred-y.T)
    w=w-alpha*(dw)
    b=b-alpha*db
    return b,w
```

```python
#Main
alpha=0.0005 #larning rate between 0 and 1 
cost=[]
b=1 #intercept
m=x.shape[0]
for i in range(2000000):
    pred=hypo(w,x,b) #predicton
    er=error(pred,y,m)
    b,w=gradient(pred,y,m,alpha,w,b)
    if i%100000==0:
        print(er)
        cost.append(er)
```

```python
plt.plot(cost)
plt.ylabel('cost')
plt.xlabel('iterations')
plt.title('Cost reduction over time')
plt.show()
```
