```python
df = pd.read_csv("/kaggle/input/the-human-freedom-index/hfi_cc_2020.csv")
```

```python
lista_cuantitativos = list()
lista_cualitativos = list()
for i,j in zip(df.dtypes.index, df.dtypes):
    if j == 'float64':
        lista_cuantitativos.append(i)
    else:
        lista_cualitativos.append(i)
```

```python
dicionario_rangos = dict()
for i in range(len(lista_cuantitativos)):
    dicionario_rangos[lista_cuantitativos[i]]=df[lista_cuantitativos[i]].max()-df[lista_cuantitativos[i]].min()
```

```python
#Obtén la moda o promedio dependiendo el caso para cada atributo.
import scipy
from scipy import stats 

dicionario_mp = dict()
for i,j in zip(df.dtypes.index, df.dtypes):
    if j == 'float64':
        dicionario_mp[i]=np.mean(df[i])
    else:
        dicionario_mp[i]=stats.mode(df[i])[0][0]
```

```python
#Obtén el rango, varianza, desviación estandar para los atributos cuantitativos.
dicionario_rangos = dict()
for i in range(len(lista_cuantitativos)):
    dicionario_rangos[lista_cuantitativos[i]]=df[lista_cuantitativos[i]].max()-df[lista_cuantitativos[i]].min()

dicionario_varianza = dict()
for i in range(len(lista_cuantitativos)):
    dicionario_varianza[lista_cuantitativos[i]]=np.var(df[lista_cuantitativos[i]])

dicionario_stdev = dict()
for i in range(len(lista_cuantitativos)):
    dicionario_stdev[lista_cuantitativos[i]]=np.std(df[lista_cuantitativos[i]])
```

Calcula el porcentaje de datos dentro de 1 desviación standard y 
dentro de 2 desviaciones standard

```python
dicionario_1stdev = dict()


for i in range(len(lista_cuantitativos)):
    
    x1 =(df[lista_cuantitativos[i]] <= (dicionario_mp[lista_cuantitativos[i]] + dicionario_stdev[lista_cuantitativos[i]])) & (df[lista_cuantitativos[i]] >= (dicionario_mp[lista_cuantitativos[i]] - dicionario_stdev[lista_cuantitativos[i]]))
        
    dicionario_1stdev[lista_cuantitativos[i]]=pd.DataFrame(x1.value_counts()).loc[True][0]/pd.DataFrame(x1.value_counts()).sum()[0]
    
```

```python
dicionario_2stdev = dict()


for i in range(len(lista_cuantitativos)):
    
    x2 =(df[lista_cuantitativos[i]] <= (dicionario_mp[lista_cuantitativos[i]] + (2 * dicionario_stdev[lista_cuantitativos[i]]))) & (df[lista_cuantitativos[i]] >= (dicionario_mp[lista_cuantitativos[i]] - (2 * dicionario_stdev[lista_cuantitativos[i]])))
        
    dicionario_2stdev[lista_cuantitativos[i]]=pd.DataFrame(x2.value_counts()).loc[True][0]/pd.DataFrame(x2.value_counts()).sum()[0]
    
dicionario_2stdev
```
