## Load Test data , Data wrngling

```python
df = pd.read_csv('/kaggle/input/bigmart-sales-data/Train.csv')
df['Item_Fat_Content'] = df['Item_Fat_Content'].apply(lambda x: 'Low Fat' if x in ['low fat', 'LF'] else x)
df['Item_Fat_Content'] = df['Item_Fat_Content'].apply(lambda x: 'Regular' if x == 'reg' else x)
```

## Quick visualization

```python
def myhist (df , xvar , ax , huevar = None , multiplev = 'layer' 
          , elementv = 'poly'):
  sns.histplot(df , x=xvar , ax=ax , hue = huevar , multiple = multiplev,
               element  = elementv)
  ax.set_xlabel(f'distribution of {xvar}'.title() , fontsize = 10)
  if huevar:
    ax.set_title(f'distrbution of {xvar} by {huevar}'.title(), fontsize = 14)
  else:
    ax.set_title(f'distrbution of {xvar}'.title(), fontsize = 14)



# create list for the qauntative virables
Quantative_variables = ['Item_Weight' , 'Item_Visibility' , 
                        'Item_MRP'  , 
                        'Item_Outlet_Sales']
```

```python
sns.set_theme()
fig , ax = plt.subplots(nrows = 2 , ncols = 2 , figsize = [12,12])
for axes , x in zip (ax.flatten() , Quantative_variables):
  myhist(df , x , axes , huevar = 'Item_Fat_Content')
```

<center><b>Item Visibility is highly left skewed <br>same for item _outlet_sales

```python
fig , ax = plt.subplots(nrows = 2 , ncols = 2 , figsize = [12,12])
for axes , x in zip (ax.flatten() , Quantative_variables):
  myhist(df , x , axes , huevar = 'Outlet_Size')
```

<center><b> Item_outlet_sales: Meduim in higher range than the high <br>
    Item_Visiability : Small out lit size take more range

```python
fig , ax = plt.subplots(nrows = 2 , ncols = 2 , figsize = [12,12])
for axes , x in zip (ax.flatten() , Quantative_variables):
  myhist(df , x , axes , huevar = 'Item_Type')
```

```python
sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})
# Initialize the FacetGrid object
pal = sns.cubehelix_palette(10, rot=-.25, light=.7)
g = sns.FacetGrid(df, row="Item_Type", hue="Item_Type", aspect=20, height=.7, palette=pal);

# Draw the densities in a few steps
g.map(sns.kdeplot, "Item_MRP",
      bw_adjust=.5, clip_on=False,
      fill=True, alpha=1, linewidth=1.5)
g.map(sns.kdeplot, "Item_MRP", clip_on=False, color="w", lw=2, bw_adjust=.5)
g.map(plt.axhline, y=0, lw=2, clip_on=False)


# Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, fontweight="bold", color=color,
            ha="left", va="center", transform=ax.transAxes)


g.map(label, "Item_MRP")

# Set the subplots to overlap
g.fig.subplots_adjust(hspace=-.25)

# Remove axes details that don't play well with overlap
g.set_titles("")
g.set(yticks=[])
g.despine(bottom=True, left=True)


```
