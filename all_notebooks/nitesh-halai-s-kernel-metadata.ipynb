```python
KernelVotes = pd.read_csv('/kaggle/input/meta-kaggle/KernelVotes.csv')
```

```python
KernelVersionKernelSources = pd.read_csv('/kaggle/input/meta-kaggle/KernelVersionKernelSources.csv')
```

```python
KernelVotes = KernelVotes.merge(KernelVersionKernelSources, how='inner', on='KernelVersionId')
```

```python
list_of_kernels = list(kernels[kernels['AuthorUserId'] == 1162156]['FirstKernelVersionId'])
list_of_kernels
```

```python
def is_my_kernel(SourceKernelVersionId):
    if SourceKernelVersionId in KernelVotes['SourceKernelVersionId']:
        return 'Nitesh'
    else:
        return 'Others'
```

```python
KernelVotes['SourceKernelVersionId'] = KernelVotes['SourceKernelVersionId'].astype('float')
```

```python
KernelVotes['Creator'] = KernelVotes['SourceKernelVersionId'].apply(is_my_kernel)
```
