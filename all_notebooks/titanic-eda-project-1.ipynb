```python
# Let's divide the whose thing in 3 categoye - Male, Female, Child
```

```python
# We will use the apply function to sort this out.
def male_female_child(passenger):
    age,sex = passenger
    if age <16 :
        return "Child"
    else:
        return sex
    
titanic_df['Pessanger_Type'] = titanic_df[['Age','Sex']].apply(male_female_child,axis=1)
```
