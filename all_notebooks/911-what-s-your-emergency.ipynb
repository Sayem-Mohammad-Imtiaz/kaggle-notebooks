```python
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])

df
```

```python
#countplot of reasons

sns.countplot(x='Reason',data=df)
plt.title ('Main Reasons for 911 calls');
```

```python
df['timeStamp'] =pd.to_datetime(df['timeStamp'])
```

```python
time = df['timeStamp'].iloc[0]
```

```python
time.year
```

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Day of week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
```

```python
df.head()
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'}
df['Day of week'] = df['Day of week'].map(dmap)
```

```python
#Friday has the most number of calls

sns.countplot(x='Day of week', data=df,hue='Reason')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.title ('Day with the most number of calls');
```

```python
sns.countplot(x='Month', data=df,hue='Reason')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
plt.title('Month with the most number of calls and its Reasons');
```

```python
df['Year'] = df['timeStamp'].apply(lambda y:y.year)
df['Year'].head()
```

```python
year = df.groupby('Year').count()
year.head()
sns.lineplot(data=year.reset_index(), x='Year', y='lat')
plt.title('Year and latitude with most number of calls');
```

```python
df['hour'] = df['timeStamp'].apply(lambda i: i.hour)
df['hour'].head()
```

```python
hourOfDay = df.groupby('hour').count()
sns.countplot(df['hour'])
plt.title('Hours with most number of calls');
```

```python
#word cloud
from wordcloud import WordCloud, ImageColorGenerator
text = " ".join(str(each) for each in df.title)
# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=200,colormap='Set3', background_color="black").generate(text)
plt.figure(figsize=(10,6))
plt.figure(figsize=(15,10))
# Display the generated image:
plt.imshow(wordcloud, interpolation='Bilinear')
plt.axis("off")
plt.figure(1,figsize=(12, 12))
plt.show()
```
