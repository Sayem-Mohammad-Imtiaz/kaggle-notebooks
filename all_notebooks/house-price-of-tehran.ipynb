```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
Hpriceframe = pd.read_csv('../input/house-price-tehran-iran/housePrice.csv')
```

# **next step**
#### Address feature is a importent feature but 
#### but that is categorical variable and we must
#### convert it to continues variable, so we do that
## you can see the list of the address at the end of the notebook

```python
Hpriceframe['CategoricalAddress'] = Hpriceframe['Address']
Address = pd.unique(Hpriceframe['Address'])
Address = Address.tolist()
AddressDict = {}
for i in Address:
    AddressDict[i] = Address.index(i)

Hpriceframe = Hpriceframe.replace({'Address':AddressDict})
Hpriceframe.head()
```

```python
sns.heatmap(Hpriceframe.corr())
```

```python
X = Hpriceframe.drop(['Price','Price(USD)','Area','CategoricalAddress'],axis=1)
y = Hpriceframe['Price(USD)']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
pd.DataFrame(model.coef_,X.columns,columns=['Coeficient'])
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn import metrics
MAE = metrics.mean_absolute_error(y_test,y_pred)
MSE = metrics.mean_squared_error(y_test,y_pred)
RMSE = np.sqrt(MSE)
```

```python
test_residuals = y_test-y_pred
sns.displot(test_residuals, bins=45,kde=True)
```
