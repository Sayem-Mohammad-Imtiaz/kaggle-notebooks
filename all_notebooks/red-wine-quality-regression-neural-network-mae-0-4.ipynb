```python
wine_data = pd.read_csv('../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv').astype('float32')
wine_data.head()
```

```python
wine_data.shape, type(wine_data), tf.size(wine_data), wine_data.dtypes
```

```python
x = wine_data.drop(['quality'], axis=1)
x.head()
```

```python
x.columns
```

```python
y = wine_data['quality']
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
ct = make_column_transformer(
  (MinMaxScaler(), ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']),
)

ct.fit(x_train)
```

```python
x_train_normalized = ct.transform(x_train)
x_test_normalized = ct.transform(x_test)
```

```python
model = tf.keras.Sequential([
  tf.keras.layers.Dense(100, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(100, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(100, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(12, activation='relu'),
  tf.keras.layers.Dropout(0.1),
  tf.keras.layers.Dense(1),
])

model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss=tf.keras.losses.mae, metrics=['mae'])

history = model.fit(x_train_normalized, y_train, epochs=700, verbose=0)
```

```python
model.evaluate(x_test_normalized, y_test)
```

**The MAE (Mean Absolute Error) of the testing data is 0.4135.**
