```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns 

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv(r'../input/crime-in-india-2019/Corruption 2019.csv')
```

```python
## if there is no unique value i will remove it
df['Category'].unique()
```

```python
df['Category'].value_counts()
```

```python
## dropping the null values
df.drop(['Unnamed: 11','Unnamed: 12','Unnamed: 13','Unnamed: 14'], axis = 1,inplace=True)
```

```python
## let's do some analysis
df1 = df[['State/UT ','Total cases']].sort_values(by='Total cases', ascending=False)
df1.head()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(x= 'State/UT ', y='Total cases', data = df1)
plt.xticks(rotation=60);
```

```python
df2 = df[['State/UT ','Cases Charge-sheeted']].sort_values(by='Cases Charge-sheeted', ascending=False)
df2.head()
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(x= 'State/UT ', y='Cases Charge-sheeted', data = df2)
plt.xticks(rotation=60);
```

```python
## is there a difference between the state and the UT ? 
df.groupby('Category')['Total cases'].mean()
```

the avg in the state is bigger

```python
plt.figure(figsize=(8,6))
sns.barplot(x = df.groupby('Category')['Total cases'].mean().index, y = df.groupby('Category')['Total cases'].mean(), data =df)
```

```python
df3 = df[['State/UT ','Cases With-drawn by the Govt']].sort_values(by='Cases With-drawn by the Govt', ascending=False)
df3.head()
```

```python
df['charge_ratio'] = df['Total cases'] / df['Cases Charge-sheeted']
```

```python
df['charge_ratio'].head()
```

```python
plt.figure(figsize=(18,6))
sns.barplot(x='State/UT ', y = 'charge_ratio', data = df)
plt.xticks(rotation = 60);
```

```python
plt.figure(figsize=(16,8))
sns.heatmap(df.corr(), annot=True)
```
