```python
fig,axes=plt.subplots(1,4,figsize=(18,10))
imp=['battery_power','px_height','px_width','ram']
k=0
for i in range(4):
    
        sns.boxplot(ax=axes[i],x="price_range",y=imp[k],data=train)
        k+=1
```

```python
fig,axes=plt.subplots(2,3,figsize=(18,10))
cat=['blue','dual_sim','four_g','three_g','touch_screen','wifi']
k=0
for i in range(2):
    for j in range(3):
        sns.boxplot(ax=axes[i,j],x="price_range",y="ram",data=train,hue=cat[k])
        k+=1
```
