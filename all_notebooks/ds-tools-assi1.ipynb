```python
x1 = np.linspace(0,5,50)
y1 = np.linspace(0,5,90)
x,y= np.meshgrid(x1,y1)
z = np.cos(x*y*100)

ax1=plt.axes(projection='3d')
ax1.contour3D(x,y,z,cmap='Greys')
```

```python
ax2=plt.axes(projection='3d')
ax2.contour3D(x,y,z)
plt.savefig('colored3d.png')
```

```python
table1= { 'name':['Mary','Joseph'],
         'drink':['wine','beer']}
table2= {'name':['Peter','Paul','Mary'],
         'food':['fesh','beans','bread']}
```

```python
df1=pd.DataFrame(table1)
df2=pd.DataFrame(table2)
print(df1)
print(df2)
```

```python
inner=pd.merge(df1,df2,on='name',how='inner')
inner
```

```python
outer=pd.merge(df1,df2,on='name',how='outer')
outer
```
