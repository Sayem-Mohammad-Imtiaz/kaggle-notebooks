# Create global map for covid vacctination (Percentage of fully vaccinated population

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv', header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
print(df.columns)
fig = px.choropleth(df,locations ='country',locationmode='country names',color='people_vaccinated_per_hundred', hover_name='country', animation_frame='date', color_continuous_scale='rainbow', range_color=(0.0,50.0))
fig.update_layout(title_text = 'Percentage of country total fully vaccinated population')
fig.show()

plotly.offline.plot(fig,filename ='vaccinations.html')
      
```

# Vaccination Progress Analysis

By using above map we could see which country that has better progress in term of vaciccination rate accross the globe. From the recent date updated in 31 May 2021, North American region was marked with orange color in the map which means the region has the highest vaccination rate compared with other region, meanwhile africa region is showing the lowest vaccination rate. The lowest vaccination rate mostly occured at lower to middle income regoion/country where it verified by recent news that mentioned most of the developing and lower income country are struggling to obtain vaccines shares. 

the most intersting fact is Rusia, despite the country ability to produced their own vaccines, the vaccination rate is still low. Based on several news information low vaccination rate was triggred by lack of vaccine supply in this country.

# Global map of earthquake occurance

```python
df = pd.read_csv('../input/earthquake-dataset/query.csv', header=0)
print(df.columns)
df.index = pd.to_datetime(df['time'])
df['time']=df.index.strftime('%Y-%m-%d %H:00:00')
df=df.sort_index()
fig = px.scatter_geo(df,lat='latitude', lon='longitude', color='mag',animation_frame='time', color_continuous_scale='rainbow', range_color=(1.0,8.0))
fig.show()
plotly.offline.plot(fig,filename ='earthquake.html')
```

# Analysis and new related Info

Above model was shown the record of earthquake data acrros the globe from period May to June 2021 with magnitude range 1.0 m to 8.0 mag,detail as follows :
1. Small earthquakes with magnitude range from 1 - 5 mag frequently hit the area at North-west part of America and area that bordered with Pacific North-West area. The frequent earthquake occurred in this area were correlated to the geological composition of this area, western United States lies along the boundaries of major tectonic plates that make up the Earth’s crust—the North American Plate and the ocean plates to the west. These plates are moving against each other, breaking up the crust along many faults like the San Andreas Fault. Faults in the East are less active and lie entirely within the North American Plate. Different case with Eastern part of America, the magnitude of the earthquake in this area can be twice of the western part, but the frequency is relatively rare.
