```python
plt.figure(figsize=(13,10))
sns.heatmap(df.corr(),annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

We can see high correlation (0.81) Pelvic incidence and Sacral slope on chart.
 What is correlation? Correlation is a mutual relationship or connection between two or more things. A correlation coefficient is a way to put a value to the relationship. Correlation coefficients have a value of between -1 and 1. A “0” means there is no relationship between the variables at all, while -1 or 1 means that there is a perfect negative or positive correlation.

```python
x=df.pelvic_incidence.values.reshape(-1,1)
y=df.sacral_slope.values.reshape(-1,1)
plt.figure(figsize=(8,8))
plt.scatter(x,y)
plt.title("Relation of Pelvic incidence and Sacral slope")
plt.xlabel("Pelvic incidence")
plt.ylabel("Sacral slope")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression as lr
linreg=lr()
linreg.fit(x,y)
```

```python
array=np.arange(min(x),max(x),0.001).reshape(-1,1)
yhead=linreg.predict(array)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(x,y,label="Samples")
plt.plot(array,yhead,label="Predict",color="red")
plt.title("Relation of Pelvic incidence and Sacral slope")
plt.xlabel("Pelvic incidence")
plt.ylabel("Sacral slope")
plt.legend()
plt.show()
```

```python
print("R-Squared score: ",linreg.score(x, y))
```

What Is R-Squared? R-squared (R2) is a statistical measure that represents the proportion of the variance for a dependent variable that's explained by an independent variable or variables in a regression model.R-Squared score of this regression model is 0.67.
