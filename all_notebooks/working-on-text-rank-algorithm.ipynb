```python
#Breaking text into individual sentences-

from nltk.tokenize import sent_tokenize
sentences=[]

for s in df['article_text']:
    sentences.append(sent_tokenize(s))




sentences=[y for x in sentences for y in x] #flattening list
#print(sentences)
```

```python
!wget http://nlp.stanford.edu/data/glove.6B.zip
!unzip glove*.zip
```

```python
#Extracting word vectors
word_embedddings={}

f=open("glove.6B.100d.txt",encoding='utf-8')
for line in f:
    values = line.split()
    word=values[0]
    coefs=np.asarray(values[1:],dtype='float32')
    word_embedddings[word]=coefs

f.close()
```

```python
len(word_embedddings)
```

```python
# remove punctuations, numbers and special characters
clean_sentences = pd.Series(sentences).str.replace("[^a-zA-Z]", " ")

# make alphabets lowercase
clean_sentences = [s.lower() for s in clean_sentences]
```

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

```python
# function to remove stopwords
def remove_stopwords(sen):
    sen_new = " ".join([i for i in sen if i not in stop_words])
    return sen_new
```

```python
# remove stopwords from the sentences
clean_sentences = [remove_stopwords(r.split()) for r in clean_sentences]
```

```python
sentence_vectors = []
for i in clean_sentences:
    if len(i) != 0:
        v = sum([word_embedddings.get(w, np.zeros((100,))) for w in i.split()])/(len(i.split())+0.001)
    else:
        v = np.zeros((100,))
    sentence_vectors.append(v)
```

```python
sim_mat=np.zeros((len(sentence_vectors),len(sentence_vectors)))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

for i in range(len(sentence_vectors)):
    for j in range(len(sentence_vectors)):
        sim_mat[i,j]=cosine_similarity(sentence_vectors[i].reshape(1,100),sentence_vectors[j].reshape(1,100))[0,0]
```

```python
import networkx as nx

nx_graph= nx.from_numpy_array(sim_mat)
scores=nx.pagerank(nx_graph)
```

```python
ranked_sentences=sorted(((scores[i],s) for i,s in enumerate(sentences)),reverse=True)
```
