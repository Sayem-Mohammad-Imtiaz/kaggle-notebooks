```python
#character to series mapping
def charVectors():
  AMINOACIDS = {}
  amino_acids  = "AGTC"
  amino_acids = "".join(sorted(amino_acids))
  amino_acids+="N"
  for i in range(len(amino_acids)):
    AMINOACIDS[amino_acids[i]]=np.zeros(5)
    AMINOACIDS[amino_acids[i]][i] = 1
  return AMINOACIDS

#Funtion to encode each sequence
def getEncode(seqence,AMINOACIDS):
  tempList = []
  for i in seqence:
    for j in AMINOACIDS[i]:
      tempList.append(j)
  return tempList

#Function to encode given sequences
def seqEncoding(dataF,path):
  AMINOACIDS = charVectors()
  encodeInputs = []
  for i in dataF['Sequence']:
    encodeInputs.append(getEncode(i,AMINOACIDS))
  dataB = pd.concat([dataF,pd.DataFrame(encodeInputs)], axis = 1)
  dataB.to_csv(path, index=False,float_format='%.2f')
```
