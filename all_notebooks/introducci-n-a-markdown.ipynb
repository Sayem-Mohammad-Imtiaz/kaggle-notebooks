# Cursiva y negritas

Las cursivas estan representadas con un asterisco, por ejemplo: *la vida es solo una hay que aprovecharla* aparecere en cursiva.

Para negritas en con doble asterisco, por ejemplo: **Si no escalas la montaña no disfrutaras el paisaje** estará en negritas.

Si quieres que sea coursiva y negritas con triple asterisco sirve, por ejemplo: ***Anita lava la tina***

# Links

Primero vamos a crear un *inline link*. Es texto entre corchetes seguido de link entre parentesis. Por ejemplo:

[texto genérico](http://www.fciencias.unam.mx)

### Formas [de desasustarse](https://img.cinemablend.com/filter:scale/quill/7/f/0/2/6/b/7f026b808b59d72abbb9f51d94e807b0ec1069d9.jpg?mw=600) 

[Los 7 videos más curiosos... - Dross](https://www.youtube.com/watch?v=jXDjB0mYxic)

[Fotos de *gatitos*](https://dailykitten.com)


Otra forma de usarse es por medio de "referencias" en el que en un par de corchetes cuadrados sobre el cual, el primero es el texto y el segundo es el sitio referenciado.

Ejemplo:

Los [portafolios de inversiones][port monex] son importantes.

Si nos fijamos en el valor de nuestro [portafolio][port monex] nos damos cuenta que tenemos mucho dinero.

Pero [este][port monex] se puede acabar por el [teorema fundamental del cálculo][teo fund]

[teo fund]: https://es.wikipedia.org/wiki/Teorema_fundamental_del_cálculo

# Citas

Solo basta con poner el símbolo de >, si queremos hacer un bloque necesitan ser más >.

Por ejemplo:

Como una vez dijo un sabio filósofo:
> Yo solo se que no se nada.

Veamos como esto no es parte de la cita. Ahora demos credito a quien la dijo:
> Yo solo se que no se nada - *Socrates*

Ahora una cita más larga:

>Entretanto, el hermano menor de Matilde, Guillermo Adelin, murió en el desastre del Barco Blanco en 1120, lo que dejó a Inglaterra en una inminente crisis de sucesión. 
>
>
>Después de la muerte de Enrique V, fue llamada a Normandía por su padre, quien hizo los arreglos para que se casara con Godofredo de Anjou y así formara una alianza para proteger sus fronteras meridionales. 

*¿Que pasa si se quita el >?*
