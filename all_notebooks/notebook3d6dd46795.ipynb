```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv('/kaggle/input/machine-learning-for-diabetes-with-python/diabetes_data.csv')
data.head()
```

```python
Input = data.drop("Outcome",axis = 1)
Output = data["Outcome"]
```

```python
Input = (Input-Input.min())/(Input.max()-Input.min())
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(Input,Output,test_size = 0.3,random_state = 42)
```

```python
lr = LogisticRegression(max_iter = 200)
lr.fit(X_train,Y_train)
lr_score = lr.score(X_test,Y_test)
Y_pred = lr.predict(X_test)
print(confusion_matrix(Y_test,Y_pred))
df = pd.DataFrame({"Actual":Y_test,"Predicted":Y_pred})
df.head()
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,Y_train)
dt_score = dt.score(X_test,Y_test)
Y_pred = dt.predict(X_test)
print(confusion_matrix(Y_test,Y_pred))
df = pd.DataFrame({"Actual":Y_test,"Predicted":Y_pred})
df.head()
```

```python
rf = RandomForestClassifier()
rf.fit(X_train,Y_train)
rf_score = rf.score(X_test,Y_test)
Y_pred = rf.predict(X_test)
print(confusion_matrix(Y_test,Y_pred))
df = pd.DataFrame({"Actual":Y_test,"Predicted":Y_pred})
df.head()
```

```python
print("Logistic Regression Accuracy on Test Data",str(lr_score*100))
print("Decision Tree Classifier Accuracy on Test Data",str(dt_score*100))
print("Random Forest Classifier Accuracy on Test Data",str(rf_score*100))

```

```python
accuracy = [lr_score,dt_score,rf_score]
models = ["Logistic Regression","Decision Tree Classifier","Random Forest Classifer"]
plt.figure(figsize = (5,5))
plt.yticks(np.arange(len(accuracy)),models)
plt.xlabel("Accuacy")
plt.ylabel("Algorithms")
plt.title("Accuracies of Different Algorithms")
plt.barh(np.arange(len(accuracy)),accuracy)
```

```python

gb = GradientBoostingClassifier(random_state=0)
parameters={
    'learning_rate': [0.05, 0.1, 0.5],
    'max_features' : [0.5, 1],
    'max_depth' : [3, 4, 5]
}
gridsearch=GridSearchCV(gb,parameters,cv=100,scoring='roc_auc')
gridsearch.fit(X_train,Y_train)
print(gridsearch.best_params_)
print(gridsearch.best_score_)
```
