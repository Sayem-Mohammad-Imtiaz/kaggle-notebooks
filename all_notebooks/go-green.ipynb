```python
import pandas as pd 
import matplotlib
import matplotlib.pyplot as plt
import numpy as np 
import seaborn as sns
```

```python
forest_data = pd.read_csv('../input/deforestation-and-forest-loss/annual-change-forest-area.csv')
forest_data.head(10)
```

```python
forest_data.shape
```

```python
forest_data.isna().sum()
```

```python
forest_data.iloc[136:140,1] = "FO"
forest_data.iloc[143:147,1] = "GF"
forest_data.rename(columns={"Entity" : "Country"}, inplace=True)
forest_data.isna().sum()
```

```python
forest_data['Country'].value_counts()
```

```python
forest_data_countries = forest_data.loc[(forest_data['Country'] == 'Mozambique') | (forest_data['Country'] == 'Netherlands') | (forest_data['Country'] == 'Netherlands') |
(forest_data['Country'] == 'India') | (forest_data['Country'] == 'South Korea')]
           
forest_data_countries
```

```python
plt.title('')
sns.lineplot(x='Year', y='Net forest conversion', data=forest_data_countries, sort=False, color='purple')
```

```python
forest_data_NL = forest_data.loc[(forest_data['Country'] == 'Netherlands')]
forest_data_NL
```

```python
forest_data_IN = forest_data.loc[(forest_data['Country'] == 'India')]
forest_data_IN
```

```python
forest_data_MB = forest_data.loc[(forest_data['Country'] == 'Mozambique')]
forest_data_MB
```

```python
forest_data_SK = forest_data.loc[(forest_data['Country'] == 'South Korea')]
forest_data_SK
```

```python
sns.lineplot(x='Year', y='Net forest conversion', data=forest_data_SK, sort=False, color='purple')
```

```python
plt.plot( 'Year', 'Net forest conversion', data=forest_data_NL, marker='', color='red', linewidth=2, label='Netherlands')
plt.plot( 'Year', 'Net forest conversion', data=forest_data_SK, marker='', color='yellow', linewidth=2, label='South Korea')
plt.legend()
plt.title('Year and Net forest conversion')
plt.show()
```
