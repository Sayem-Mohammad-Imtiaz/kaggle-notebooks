<h2> Start Time Analysis

<h3> Question: What time of the day are accidents most freqent in ?

```python
data.Start_Time
```

```python
#Converting Start_Time to Date Time
data.Start_Time=pd.to_datetime(data.Start_Time)
```

```python
data.Start_Time[0]
```

```python
#Extracting Hour from TimeStamp
data.Start_Time.dt.hour
```

```python
plt.figure(figsize=(16,7))

sns.distplot(data.Start_Time.dt.hour,kde=False,bins=24,norm_hist=True,color=['#67301a'])
```

<h3>Conclusion: Most of the accidents occurs during 7 a.m.-10 a.m. and 3 p.m.-9 p.m.

<h3>Question: Which days of the week have the most accidents ?And what is the distribution by the hour on weekdays and weekends ?

```python
#Analyzing the trend during the week.
plt.figure(figsize=(10,7))
sns.distplot(data.Start_Time.dt.dayofweek,kde=False,bins=7,norm_hist=True,color='#114060')
```

<h3> Most of the accidents occurs during the weekdays.

```python
#Extracting Time range for the Weekends
sundays_start_time=data.Start_Time[data.Start_Time.dt.dayofweek==6]
sundays_start_time
```

```python
#Plotting The Accidents Variation On Sunday
plt.figure(figsize=(16,7))
sns.distplot(sundays_start_time.dt.hour,kde=False,norm_hist=True,bins=24,color='#bec398')
```

```python
#Extracting Time range for the Weekends
mondays_start_time=data.Start_Time[data.Start_Time.dt.dayofweek==0]
mondays_start_time
```

```python
#Plotting The Accidents Variation On Monday
plt.figure(figsize=(16,7))
sns.distplot(mondays_start_time.dt.hour,kde=False,norm_hist=True,bins=24)
```

<h3>Question: Which months have the most accidents ?

```python
#Plotting Variation Of accidents by month
plt.figure(figsize=(16,7))
sns.distplot(data.Start_Time.dt.month,kde=False,bins=12,norm_hist=True,color='#63b6ba')
```

<h3> Insight: Since the number of accidents are unusally very low for Starting Months of the Year.
    This means that there might be lost of data during the initial years of the data collection

<h3> Question: Is there any data discrepency because 1306 states have only 1 accident ? What is the trend in the accidents over the years ?

```python
#Extracting The Data For the year 2019
data_2019=data[data.Start_Time.dt.year==2019]
```

```python
data_2019
```

```python
plt.figure(figsize=(16,7))
sns.distplot(data_2019.Start_Time.dt.month,kde=False,bins=12,norm_hist=True,color='#6942c2')
```

```python
#Extracting The Data For the year 2016 and comparing it with 2019
data_2016=data[data.Start_Time.dt.year==2016]
```

```python
plt.figure(figsize=(16,7))
sns.distplot(data_2016.Start_Time.dt.month,kde=False,bins=12,norm_hist=True,color='#6b594b')
```

<h3>Conclusion: There is loss of data during the initial months of the year 2016
