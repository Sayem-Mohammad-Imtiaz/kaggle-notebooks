```python
np.set_printoptions(suppress=True, precision = 2)
df = pd.read_csv('../input/nba-games/NBA all data.csv')
```

```python
df.info()
df['point_diff'] = df['pts'] - df ['opp_pts']
```

```python
df_10 = df[df['year_id'] == 2010]
df_14 = df[df['year_id'] == 2014]
```

```python
knick_pts = df_10.pts[df_10['fran_id'] == 'Knicks']
nets_pts = df_10.pts[df_10['fran_id'] == 'Nets']
knick_pts_mean = np.mean(knick_pts)
nets_pts_mean = np.mean(nets_pts)
diff_means_2010 = knick_pts_mean - nets_pts_mean
```

```python
figure(figsize = (8, 8), dpi=80)
plt.xlabel('Points')
plt.ylabel('Times')
plt.title('Knick Vs Neats Games', fontsize = 16)
plt.hist(knick_pts, edgecolor = 'black', density = True, alpha = 0.5, color = '#008080', label = 'Knick')
plt.hist(nets_pts, edgecolor = 'black', density = True, alpha = 0.5, color = '#9400d3', label = 'Neats')
plt.legend()
plt.show()
```

```python
points_2010 = np.mean(df_10.pts)
points_2014 = np.mean(df_14.pts)
diff_means_years = points_2014 - points_2010
```

```python
figure(figsize = (8, 8), dpi=80)
plt.xlabel('Points Throught Years')
plt.ylabel('Times')
plt.title('Points ', fontsize = 16)
plt.hist(df_10.pts, edgecolor = 'black', density = True, alpha = 0.5, color = '#D41313', label = '2010')
plt.hist(df_14.pts, edgecolor = 'black', density = True, alpha = 0.5, color = '#1382D4', label = '2014')
plt.legend()
plt.show()
```

```python
figure(figsize = (12, 12), dpi=100)
sns.set_style('darkgrid')
sns.set_palette('Set2')
plt.title('Points by Year')
sns.boxplot(y = df_10.pts, x = df_10.fran_id)
plt.xlabel('Teams')
plt.ylabel('Points')
plt.xticks(rotation = 90)

plt.show()
```

```python
print('Frequency')
location_result_freq = pd.crosstab(df_10.game_result, df_10.game_location)
print(location_result_freq)
print('Proportions')
location_result_proportion = location_result_freq / len(df_10)
print(location_result_proportion)
```

> # The 'Game Location' and 'Game results' are associated

```python
chi2, pval, dof, expected = chi2_contingency(location_result_proportion)
print(expected)
print(chi2)
```

> # We can identify covariance get in points by the NBA's teams and point difference that it's the marginal covariance.

```python
cov_forecast_ptsdiff = np.cov(df_10['forecast'], df_10['point_diff'])
print(cov_forecast_ptsdiff)
```

```python
correlation_forecast_ptsdiff, p = pearsonr(df_10['forecast'], df_10['point_diff'])
print(correlation_forecast_ptsdiff)
```

```python
figure(figsize = (7, 7), dpi=80)
plt.scatter(x = df_10['forecast'], y = df_10['point_diff'])
plt.xlabel('Forecasted Win Prob.')
plt.ylabel('Point Differential')
plt.tight_layout()
plt.show()
```
