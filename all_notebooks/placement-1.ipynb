```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.linear_model import LinearRegression
```

```python
#import the dataset
data = pd.read_csv('/kaggle/input/factors-affecting-campus-placement/Placement_Data_Full_Class.csv')
data.head()
```

### Q1.

Develop an estimated multiple linear regression equation with `mba_p` as response variable and `ssc_p` & `hsc_p` as the two predictor variables. 

- Interpret the regression coefficients and check whether they are significant based on the summary output 

```python
data.shape
```

```python
# the predictor variables
data_features = ['ssc_p','hsc_p']
X = data[data_features]
```

```python
#the response variable
y = data.mba_p
```

```python
model = LinearRegression()
model.fit(X,y)
r2 = model.score(X,y)
```

```python
#print the coefficients
print(f'alpha = {model.intercept_}')
print(f'Betas = {model.coef_}')
```

```python
import statsmodels.api as sm
mod = sm.OLS(y,X)
fitmodel = mod.fit()
fitmodel.pvalues
```

#### Interpreting coefficients

`mba_p = 44.61 + 0.15 ssc_p + 0.11 hsc_p`

The coefficients show that there is a positive relationship between the response variable and predictor variables. 

The two predictor variables are also statistically significant since their p-values are less than the usual significance level of 0.05.
