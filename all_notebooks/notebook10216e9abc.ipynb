# **read the two file ****

```python
data = pd.read_csv('/kaggle/input/train-data/data.csv')
test_data = pd.read_csv('/kaggle/input/test-data/test_data.csv')
```

# ***Chabge the value of Male by 0 and female by 1***

```python
data.loc[data['Genre'] == 'Male', 'Genre'] = 0
data.loc[data['Genre'] == 'Female', 'Genre'] = 1
```

# ***the data is clean so we will start classification by K-Means Clustering***

```python
train = data.iloc[:,3:6].values
```

```python
y=data.Class
ind=['Genre','Annual Income (k$)','Age']
x=data[ind]
plt.hist(y)
```

```python
new_data_test = test_data[["Genre","Age","Annual Income (k$)"]].copy()
new_data_test.loc[new_data_test['Genre'] == 'Male', 'Genre'] = 0
new_data_test.loc[new_data_test['Genre'] == 'Female', 'Genre'] = 1
```

# **the classification code ****

```python
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(x, y)
Y_pred = random_forest.predict(new_data_test)
```

```python
Y_pred
```

# **the data below is the data after classification****

```python
the_predection_class = pd.DataFrame(data=Y_pred, columns=["Class"])
result = pd.concat([test_data, the_predection_class], axis=1)
result
```
