```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import math
import seaborn as sns
```

```python
# Importing the dataset
dataset = pd.read_csv('../input/ecommerce-customers/Ecommerce Customers.csv')
x=dataset.iloc[:,3:7].values
y=dataset.iloc[:,-1].values
```

```python
#spliting data into training and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#using the decision tree regrtession on dataframe
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(x_train,y_train)
```

```python
#predicting results using decision tree
Y_pred=regressor.predict(x_test)
```

```python
#using the multiple linear regression
from sklearn.linear_model import LinearRegression
linear=LinearRegression()
linear.fit(x_train,y_train)
```

```python
#predicting anuual amount spend by customers using multiple linear regression
multiple_pred=linear.predict(x_test)
```

```python
#calculating r2 score of multiple regression 
from sklearn.metrics import r2_score
r2_score(y_test,multiple_pred)
```

```python
#calculating r2 score of decision tree
from sklearn.metrics import r2_score
r2_score(y_test,Y_pred)
```
