```python
import pandas as pd
data=pd.read_csv("../input/80-cereals-nutrition-data-on-80-cereal-products/cereal.csv")
data.head()
```

```python
data['mfr'].unique()
```

```python
data['type'].unique()
```

```python
import sklearn 
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
label_encoder=preprocessing.LabelEncoder()
data['type']=label_encoder.fit_transform(data['type'])
data['mfr']=label_encoder.fit_transform(data['mfr'])
data['name']=label_encoder.fit_transform(data['name'])

```

```python
data.head()
```

```python
data.isnull().any() ####To check if there are any missing values in the data
```

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,roc_auc_score
```

```python
X=data.drop(['type'],axis=1)
y=data['type']
```

```python
X.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
import warnings
lm=LogisticRegression()
lm.fit(X_train,y_train)
```

```python
y_predict=lm.predict(X_test)
y_predict
```

```python
####To decide is the model a good fit###
confusion_matrix(y_test,y_predict)
```

```python
roc_auc_score(y_test,y_predict)
```
