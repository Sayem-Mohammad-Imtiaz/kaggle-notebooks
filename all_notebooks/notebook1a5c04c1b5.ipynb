```python
#categorizing the DISTANCE column:
def fixdistance(x):
  if (x>=1000) & (x<1333):
    return 'SHORT FLIGHT'
  if (x>=1333) & (x<2666):
    return 'MEDIUM FLIGHT'
  if (x>=2666):
    return 'LONG FLIGHT'
```
