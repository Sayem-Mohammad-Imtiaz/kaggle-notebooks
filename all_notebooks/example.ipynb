```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/russia-real-estate-20182021/all_v2.csv")
```

```python
df.info()
```

```python
f, axes = plt.subplots(1, 4)  
sns.boxplot(  y="object_type", data=df, ax=axes[0])
sns.boxplot(  y="building_type", data=df, ax=axes[1])
sns.boxplot(  y="level", data=df, ax=axes[2])
sns.boxplot(  y="levels", data=df, ax=axes[3])
plt.tight_layout()
```

```python
f, axes = plt.subplots(1, 2)  
sns.boxplot(  y="area", data=df, ax=axes[0])
sns.boxplot(  y="kitchen_area", data=df, ax=axes[1])
plt.tight_layout()
```

```python
f, axes = plt.subplots(1, 2)  
sns.boxplot(  y="rooms", data=df, ax=axes[0])
sns.boxplot(  y="price", data=df, ax=axes[1])
plt.tight_layout()
```

```python
print(df.groupby(['rooms']).size())
print(df.groupby(['level']).size())
print(df.groupby(['levels']).size())
print(df.groupby(['object_type']).size())
print(df.groupby(['building_type']).size())
```
