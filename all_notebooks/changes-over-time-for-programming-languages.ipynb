```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
dataset = pd.read_csv("/kaggle/input/most-popular-programming-languages-since-2004/Most Popular Programming Languages from 2004 to 2020 V3.csv")
dataset.head()
```

```python
dataset.shape
```

```python
dataset.dtypes
```

```python
dataset["Date"] = pd.to_datetime(dataset["Date"])
dataset.set_index("Date", inplace=True)
dataset
```

What was the most popular programming language in 2004 and which is the most popular in 2020?

```python
popular_2004 = pd.DataFrame({"Languages": dataset.iloc[0].T.index,
                            "Popularity": dataset.iloc[0].values.T})

popular_2020 = pd.DataFrame({"Languages": dataset.iloc[197].T.index,
                            "Popularity": dataset.iloc[197].values.T})
```

```python
fig, ax = plt.subplots(1,2, figsize=(18,15))
plt.suptitle(" 10 Most Popular Programming Languages",fontsize=20)

ax[0].pie(popular_2004.nlargest(10,"Popularity").Popularity.values, autopct='%1.1f%%',
            labels=popular_2004.nlargest(10,"Popularity").Languages.values, labeldistance=0.7)
ax[0].set_title("2004",fontsize=15)

ax[1].pie(popular_2020.nlargest(10,"Popularity").Popularity.values, autopct='%1.1f%%',
            labels=popular_2020.nlargest(10,"Popularity").Languages.values, labeldistance=0.7)
ax[1].set_title("2020",fontsize=15)

plt.tight_layout()
plt.show()
```

We can statistically analyze changes over time

```python
#provides the absolute changes between two index values 
dataset.diff().mean().plot(kind='bar',figsize=(10,8))
```

We see that Python was the language with the most positive change in popularity, while Java and PHP had negative changes in their popularity. 

From a statistical point of view, absolute changes are not optimal because they are depend on the scale of the time series data itself. Therefore, percentage changes are usually preferred. 

```python
#calculates the percentage change between two index values
dataset.pct_change().mean().plot(kind='bar',figsize=(10,8))
```

We now have a clear view that there have been significant increasing changes in popularity over time in various programming languages, not only in Python, but also in Scala, Rusty, Objetive-C and Groovy.
