```python
df=pd.read_csv("../input/bostonhoustingmlnd/housing.csv")
df.shape
df.head()
df.isnull().sum()
df.plot.scatter('RM','LSTAT')
df['RM'].corr(df['LSTAT'])
df.plot.scatter('RM','PTRATIO')
df['RM'].corr(df['PTRATIO'])
df['LSTAT'].corr(df['PTRATIO'])
df['RM'].corr(df['MEDV'])
df.plot.scatter('RM','MEDV')
df['LSTAT'].corr(df['MEDV'])
df.plot.scatter('LSTAT','MEDV')
df['PTRATIO'].corr(df['MEDV'])
df.plot.scatter('PTRATIO','MEDV')
df.describe()
df.shape
train=df[0:399]
test=df[400:]
x_train=train.drop('MEDV',axis=1)
y_train=train['MEDV']
x_test=test.drop('MEDV',axis=1)
y_test=test['MEDV']
from sklearn.linear_model import LinearRegression
lreg=LinearRegression()
lreg.fit(x_train,y_train)
pred=lreg.predict(x_test)
pred
lreg.score(x_train,y_train)
lreg.score(x_test,y_test)
my_median_pred=lreg.predict([[6,5,16]])
my_median_pred

```
