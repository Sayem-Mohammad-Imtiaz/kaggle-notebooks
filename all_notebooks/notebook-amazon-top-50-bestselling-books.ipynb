```python
Avg_User_Rating = ((df['User Rating'])*(df['Reviews']))/(df['Reviews'] + (df['Reviews'].min())) + (df['Reviews'].min()*df['User Rating'].mean())/(df['Reviews'] + df['User Rating'].min())
Avg_User_Rating = Avg_User_Rating
Avg_User_Rating
```
