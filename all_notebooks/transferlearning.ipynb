```python
from tensorflow.python.keras.models import Sequential
import numpy as np
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
```

```python
weights_path = '../input/inceptionv3/inception_v3_weights_tf_dim_ordering_tf_kernels_notop.h5'
num_classes = 2
my_new_model = Sequential()
my_new_model.add(InceptionV3(include_top=False,
                           pooling='avg',
                           weights='imagenet'
))

my_new_model.add(Dense(num_classes, activation = 'softmax'))
my_new_model.layers[0].trainable = False
```

```python
my_new_model.compile(optimizer = 'Adam', 
                     loss = 'binary_crossentropy',
                     metrics = ['accuracy'])
```

```python
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator
data_generator = ImageDataGenerator(preprocessing_function = None)
#rotation_range=0,width_shift_range=0.0,height_shift_range=0.0,
```

```python
train_generator = data_generator.flow_from_directory('../input/pomaranczowetestset/training',
                                                    target_size = (300, 300),
                                                    batch_size = 10,
                                                    class_mode = 'categorical')
#w zaleznosci od batch size (=12 tutaj, potem steps per epoch = 6, poniewaz mamy łącznie 72 obrazy w secie)
```

```python
validation_generator = data_generator.flow_from_directory('../input/pomaranczowetestset/validation',
                                                         target_size = (300, 300),
                                                         batch_size = 40,
                                                         class_mode = 'categorical')
#validation steps = validation data / validation batch size
```

```python
test_generator = data_generator.flow_from_directory('../input/pomaranczowetestset/test',
                                                         target_size = (300, 300),
                                                         batch_size = 20,
                                                         class_mode = 'categorical')
```

```python
my_new_model.fit_generator(train_generator,
                           steps_per_epoch = 58,
                          validation_data = validation_generator,
                          validation_steps = 1,
                          epochs = 200)
```

```python
from keras.applications.imagenet_utils import preprocess_input, decode_predictions
preds = my_new_model.predict(test_generator)
#print('Predicted:', preds)
```
