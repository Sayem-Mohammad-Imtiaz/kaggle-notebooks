```python
#!pip install pycaret
from pycaret.clustering import * 
exp = setup(df1)
kmeans = create_model("kmeans",num_clusters=4)
```

```python
plot_model(kmeans,plot="elbow")
```

```python
plot_model(kmeans)
```

```python
df2 = predict_model(kmeans, data = df1)
df2["country"]=df["country"]
df2
```
