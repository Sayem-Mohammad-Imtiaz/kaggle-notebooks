```python
data_set = pd.read_csv('/kaggle/input/years-of-experience-and-salary-dataset/Salary_Data.csv')
```

```python
data_set.head()
```

```python
%matplotlib inline
plt.xlabel('Experience')
plt.ylabel('salary')
plt.scatter(data_set.YearsExperience, data_set.Salary, color='red')
plt.show()
```

```python
x = data_set.iloc[:,0:1].values
y = data_set.iloc[:,1].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, x_test, Y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
Y_train
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(x_test, y_test)
```

```python
%matplotlib inline
plt.xlabel('Experience')
plt.ylabel('salary')
plt.title('Years vs Salary (Training model)')
plt.scatter(X_train, Y_train)
plt.plot(X_train, model.predict(X_train), color = 'red')
```

```python
%matplotlib inline
plt.xlabel('Experience')
plt.ylabel('salary')
plt.title('Years vs Salary (Testing model)')
plt.scatter(x_test, y_test, color='green')
plt.plot(x_test, model.predict(x_test), color = 'yellow')
plt.show()
```
