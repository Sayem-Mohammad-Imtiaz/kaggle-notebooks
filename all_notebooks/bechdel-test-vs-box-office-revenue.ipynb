Now for our study we will be only focusing on the clean_test coulumn and the domgross column

```python
cleantest = movies.clean_test
domgross = movies.domgross

options = ["ok","notalk","men","nowomen","dubious"]

labels = ["Passes","No Talk","Men","No Women","Dubious"]
sizes = []
count = 0
for x in options:
    for y in range(len(cleantest)):
        if(cleantest[y] == x):
            count = count + 1
    sizes.append(count)
    count = 0

separated=(.1,0,0,0,0,)
fig1, ax1 = plt.subplots()

ax1.pie(sizes, explode=separated, labels=options, autopct='%1.2f%%',
        shadow=True, startangle=90, textprops={'fontsize': 15})
ax1.axis('equal')  
plt.title("Distrobution of of the clear test throughout all movies", fontsize=25)
fig = plt.gcf()
fig.set_size_inches(8,8)
plt.show()
```

So what does each one of these mean?

1. Passes - This means that the movie has a equal or greater amount of involvement by females 
2. No Talk - This means that although there are women in the movie they do not talk throughout the entire movie
3. Men - This means that the women in the movie only talk about men and nothing else
4. Dubious - THis means that it does not pass more for some other more specific reason that is not listed above.

The pie chart above shows you that a majority of movies do not have a powerful femal charactor. But lets delve deeper. Does this relate to domestic gross or the money made from the movie in the U.S.? Is there a relationship between these two variables?
