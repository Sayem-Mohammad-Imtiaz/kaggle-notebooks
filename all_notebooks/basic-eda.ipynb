<h1>Let's find the number of players from each country</h1>

```python
org = pd.DataFrame(df.groupby('Origin', as_index=False,sort=True)['Player'].count())
org.sort_values(by=['Player'], ascending= False, inplace=True)
org.head()
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(org['Origin'], org['Player'], color ='cyan',
        width = 0.4)
 
plt.xlabel("Origin")
plt.ylabel("No.of Players")
plt.title("No.of players from each country")
plt.xticks(rotation = 90)
plt.show()
```

<h4>From this we can conclude that most players are from France, Portugal, Brazil</h4>

<h1>Let's find the price of each player</h1>

```python
pf= df[['Player','Fee(€ mln)']]
pf.sort_values(by=['Fee(€ mln)'], ascending= False, inplace=True)
pf.head()
```

```python
fig = plt.figure(figsize = (18, 5))
plt.bar(pf['Player'], pf['Fee(€ mln)'], color ='wheat',
        width = 0.4)
 
plt.xlabel("Player")
plt.ylabel("Price (in €)")
plt.title("Price of each Player in €")
plt.xticks(rotation = 90)
plt.show()
```

<h4>This tells us that Neymar is the costliest player</h4>

<h1>Lets's find the revenues of each club</h1>

```python
club = pd.DataFrame(df.groupby('From(Club)', as_index=False,sort=True)['Fee(€ mln)'].sum())
club.sort_values(by=['Fee(€ mln)'], ascending= False, inplace=True)
club.head()
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(club['From(Club)'], club['Fee(€ mln)'], color ='#0faa13',
        width = 0.4)
 
plt.xlabel("Club")
plt.ylabel("Revenue (in €)")
plt.title("Revenue generated by each club (in €)")
plt.xticks(rotation = 90)
plt.show()
```

<h4>This shows that Barcelona has generated the most revenue and Atlético Madrid, Monaco, Borussia Dortmund, Borussia Dortmund have all crossed the 300 mark</h4>

<h1>Revenue generated for each position</h1>

```python
df['Position'].unique()
```

```python
#Let's rename Striker as Forward
df['Position'] = df['Position'].apply(lambda x : x if x!='Striker' else 'Forward')
pos = pd.DataFrame(df.groupby('Position', as_index=False,sort=True)['Fee(€ mln)'].sum())
pos.sort_values(by=['Fee(€ mln)'], ascending= False, inplace=True)
pos.head()
```

```python
fig = plt.figure(figsize = (10, 5))
plt.barh(pos['Position'], pos['Fee(€ mln)'], color ='#bc9902')
 
plt.xlabel("Revenue (in €)")
plt.ylabel("Position")
plt.title("Total revenue for each position")
# plt.xticks(rotation = 90)
plt.show()
```
