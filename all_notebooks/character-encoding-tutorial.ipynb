```python
#Open the text file as a read only file
with open ("../input/character-encoding-examples/harpers_ASCII.txt",'r') as ro:
    harper = ro.read()

#Now encode the file default encoding is utf_8 so no need to specify the desired character encoding
harper_utf_8=harper.encode()

#Check to see the difference 
print("Type of file (harper):",type(harper),"\n","\n"+str(harper[:545]),"\n"+140*"_","\nType of file (harper_utf_8):",
      type(harper_utf_8),"\n","\n"+str(harper_utf_8[:545]))
```

Let's have one more example to make sure we've fully understood the process

```python
#First open the file in bytes format
with open ("../input/character-encoding-examples/yan_BIG-5.txt",'rb') as ro:
    yan_BIG_5 = ro.read()

#Decode the file using the encoding it was written with    
yan_BIG_5_decoded=yan_BIG_5.decode("BIG5")

#Now encode the file using the utf_8 
yan_BIG5_utf8=yan_BIG_5_decoded.encode()

#Check 
print(yan_BIG5_utf8[:1000])
```
