```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
import warnings
warnings.filterwarnings('ignore')
```

<h1 id="heading3">
<span style="font-size:22px;">Loading Data</span>
<a class="anchor-link" href="https://www.kaggle.com/alampalsingh/notebook004d896e8c#heading3" target ='_self'></a>
</h1>

```python
df = pd.read_csv('../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv')
```

```python
df
```

```python
df.age.describe()
```

<span style="font-size:22px;">**Skewness**

```python
_,ax = plt.subplots(3,4,figsize = (20,15))
ax = ax.flatten()
for i in range (0,df.shape[1]-1):
    sns.kdeplot(df.iloc[:,i],ax = ax[i])
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(df.corr(),annot = True,square = False,yticklabels = False,linewidth = 1,linecolor = 'black')
```

```python
_,ax = plt.subplots(3,4,figsize = (25,15))
ax = ax.flatten()
for i in range (0,df.shape[1]-1):
    sns.violinplot(x = df.DEATH_EVENT,y = df.iloc[:,i],ax = ax[i],split=True,scale="count", inner="quartile",palette="Set2")
```

```python
#Converting age in different grops
for i in range(0,df.shape[0]):
    if 40 <= df.iloc[i,0] < 58 :
        df.iloc[i,0] = 0
    elif 58 <= df.iloc[i,0] < 76 :
        df.iloc[i,0] = 1
    elif 76 <= df.iloc[i,0] < 96 :
        df.iloc[i,0] = 2
```

```python
x = df.drop(['DEATH_EVENT'],axis = 1)
```

```python
y = df.DEATH_EVENT
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

<h1 id="heading6">
<span style="font-size:22px;">Random Forest with variable random_state.</span>
<a class="anchor-link" href="https://www.kaggle.com/alampalsingh/notebook004d896e8c#heading6" target ='_self'></a>
</h1>

```python
h = []
for i in np.arange(0,100):
    rfc = RandomForestClassifier(random_state=i)
    rfc.fit(x_train,y_train)
    acc = accuracy_score(y_test,rfc.predict(x_test))
    h.append(acc)
max(h),np.mean(h)

plt.figure(figsize=(15,6))
plt.axhline(y = np.mean(h), color = 'g',linestyle = '-')
plt.plot(np.arange(0,100),h)
plt.annotate('Max',(h.index(max(h)),max(h)),xytext = (10,0.79),arrowprops=dict(arrowstyle='->', color='red',  linewidth=3, mutation_scale=30))
plt.annotate('Mean',(58,np.mean(h)),xytext = (57,0.76),arrowprops=dict(arrowstyle='->', color='red',  linewidth=3, mutation_scale=30))
```

<span style="font-size:18px;">**ðŸ™„ Huge variation isn't it.**

<h1 id="heading7">
<span style="font-size:22px;">XGBoost</span>
<a class="anchor-link" href="https://www.kaggle.com/alampalsingh/notebook004d896e8c#heading7" target ='_self'></a>
</h1>

```python
h = []
for i in np.arange(0,100):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=i)
    xgb = XGBClassifier(max_depth= 4,reg_alpha=10,learning_rate= 0.01,verbosity = 0)
    xgb.fit(x_train,y_train)
    acc = accuracy_score(y_test,xgb.predict(x_test))
    h.append(acc)
max(h),np.mean(h)

plt.figure(figsize=(15,6))
plt.axhline(y = np.mean(h), color = 'g',linestyle = '-')
plt.plot(np.arange(0,100),h)
plt.annotate('Max(91.67%)',(h.index(max(h)),max(h)),xytext = (10,0.91),arrowprops=dict(arrowstyle='->', color='red',  linewidth=3, mutation_scale=30))
plt.annotate('Mean',(60,np.mean(h)),xytext = (60,0.90),arrowprops=dict(arrowstyle='->', color='red',  linewidth=3, mutation_scale=30))
```

<h1 id="heading8">
<span style="font-size:22px;">LightGBM with variable random_state.</span>
<a class="anchor-link" href="https://www.kaggle.com/alampalsingh/notebook004d896e8c#heading8" target ='_self'></a>
</h1>


```python
h = []
for i in np.arange(0,100):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=i)
    lgb = LGBMClassifier(max_depth= 4,reg_alpha=10,learning_rate= 0.01)
    lgb.fit(x_train,y_train)
    acc = accuracy_score(y_test,lgb.predict(x_test))
    h.append(acc)
max(h),np.mean(h)

plt.figure(figsize=(15,6))
plt.axhline(y = np.mean(h), color = 'g',linestyle = '-')
plt.plot(np.arange(0,100),h)
plt.annotate('Max(96.67%)',(h.index(max(h)),max(h)),xytext = (30,0.95),arrowprops=dict(arrowstyle='->', color='red',  linewidth=3, mutation_scale=30))
plt.annotate('Mean',(60,np.mean(h)),xytext = (60,0.90),arrowprops=dict(arrowstyle='->', color='red',  linewidth=3, mutation_scale=30))
```
