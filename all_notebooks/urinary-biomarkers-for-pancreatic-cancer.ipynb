```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df_original = pd.read_csv('../input/urinary-biomarkers-for-pancreatic-cancer/Debernardi et al 2020 data.csv')
df_original.head()
```

```python
y = df_original.diagnosis
df = df_original.drop(columns=['sample_id','patient_cohort','sample_origin','diagnosis','stage','benign_sample_diagnosis'])
df.head()
```

```python
L=len(y)
for i in range(L):
    if y[i]==3:
        y[i]=1
    else:
        y[i]=0
```

```python
for i in df[2:]:
    if i!='sex':
        df[i] = df[i].fillna(df[i].mean())
df.head()
```

```python
df=pd.get_dummies(df)
df.head()
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split

clf = svm.SVC()

X_train, X_test, y_train, y_test = train_test_split(df, y, random_state=42)

SVM = clf.fit(X_train, y_train)

score = SVM.score(X_test, y_test)
print('Test set score: {}'.format(score))
```
