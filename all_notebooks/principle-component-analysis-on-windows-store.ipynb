# Cleaning the Data
* Cheking for NULL VALUES through the Matrics
* Performing dropna just to be sure if any remaining rows are left
* Changing the date column to datetime
* Creating a new column of Recency which will tell how recent was the rating
* Changing the Price column's value 'Free' to 0 for numeric datatype.

```python
msno.matrix(df)
```

* The matrix tells us that there is any major data missing from any column
* we will still perform a general dropna to be completely sure that all the data is free from any small null values if present

```python
df.dropna(inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.info()
```

```python
latest_date = df['Date'].max()

df['Recency (x days ago)'] =df['Date'] - latest_date 
df['Recency (x days ago)'] = df['Recency (x days ago)'].dt.days
```

```python
df.head()
```

```python
df.groupby(df['Price']).count()# found out that only the value 'Free' was non-numeric
df['Price'] = df['Price'].replace('Free', '0')# Changed it to numeric
```

```python
df.head()
```

# Data is now ready!

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(df[['Rating', 'No of people Rated', 'Recency (x days ago)']])
scaled_data = scaler.transform(df[['Rating', 'No of people Rated', 'Recency (x days ago)']])
```

```python
scaled_data = pd.DataFrame(scaled_data)
```

```python
from sklearn.decomposition import PCA
logmodel = PCA(n_components=2)
logmodel.fit(scaled_data)
pca_x = logmodel.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
pca_x.shape
```

```python
df_comp = pd.DataFrame(logmodel.components_, columns = ['Rating', 'No of people Rated', 'Recency (x days ago)'])
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df_comp)
```

```python
plt.figure(figsize=(10,6))
ty=sns.scatterplot(pca_x[:,0], pca_x[:,1])
sns.despine(left=True)
ty.set_title('PCA Results')
ty.set_ylabel('Second Principle Component ')
ty.set_xlabel('First Principle Component ')


```
