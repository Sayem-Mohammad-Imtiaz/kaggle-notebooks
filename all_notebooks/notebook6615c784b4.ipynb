### Converting haves and haves_not

```python
col_to_boolean= ["beans","cocoa_butter","vanilla","lecithin","salt","sugar","sweetener_without_sugar"]

def have_not_have_to_bool(x):
    if "have_not" in x:
        y=0
    else:
        y=1
    return y

for col in col_to_boolean:
    df[col]=df[col].apply(have_not_have_to_bool)
df.head()
```

# Predcitive



## OLS v1 adding only nominal variables

```python
import statsmodels.api as sm

col_for_normal=["cocoa_percent","rating"]
X=df[["cocoa_percent","counts_of_ingredients"]]
y=df[["rating"]]
X = sm.add_constant(X)

model = sm.OLS(y,X)

results = model.fit()
results.summary()
```

```python
df.head()
```

* Numerical regression= OLS
* Nominal regression = probit 
* Ordinal regression = logit

## OLS v2 adding all variables

```python
X=df[["cocoa_percent","counts_of_ingredients","beans","cocoa_butter","lecithin","salt","vanilla","sugar","sweetener_without_sugar"]]
y=df[["rating"]]

model = sm.OLS(y, X)

results = model.fit()

results.summary()
```

* If the p-value is less than 0.05, 

    - we reject the null hypothesis that there's no difference between the means and conclude that a significant difference does exist. If the p-value is larger than 0.05, we cannot conclude that a significant difference exists.03-Dec-2015

* H0:that there's no difference between the means
* H1:difference exist

## PCA

```python
from sklearn.decomposition import PCA

X=df[["cocoa_butter","sugar","vanilla","sweetener_without_sugar"]]
# X=df[["cocoa_percent","counts_of_ingredients","beans","cocoa_butter","lecithin","salt","vanilla","sugar","sweetener_without_sugar"]]


pca = PCA(n_components=2)
principalComponents = pca.fit_transform(X)

pca_df= pd.DataFrame(data = principalComponents
             , columns = ['pca1', 'pca2',], index=df.index)

pca_df["rating"]=df["rating"].round(1).astype(float)
pca_df["rating"]*=10
pca_df["rating"]=pca_df["rating"].astype(int)
display(pca_df.head())


fig = go.Figure(data=go.Scatter(
                x=pca_df["pca1"], y=pca_df["pca2"],
                mode='markers',
                hovertext=pca_df["rating"],
                marker=dict(
                    size=16,
                    color=pca_df["rating"],
                    colorscale='aggrnyl',
#                     colorscale="blues",
                    showscale=True)
                ))


fig.update_layout(title='Using PCA to reduce dimension',width=600,height=600)

fig.show()
```

## Clustering

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=6, random_state=0).fit(pca_df)
kmeans.labels_

pca_df["label"]=kmeans.labels_
display(pca_df.head())



fig = px.scatter(pca_df, x="pca1", y="pca2", color="rating",  size="rating",hover_data=["rating"], )
fig.update_layout( title="Rating", width=500, height=400 ,)
fig.show()


fig = px.scatter(pca_df, x="pca1", y="pca2", color="label",size="rating", hover_data=["rating","label"])
fig.update_layout( title="Cluster", width=500, height=400 )
fig.show()
```

## Verifying clusters

```python
kmean_df=pca_df[["rating","label"]].groupby("label").mean()
kmean_df["std"]=pca_df[["rating","label"]].groupby("label").std()["rating"]
display(kmean_df)
```

```python
import plotly.express as px
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import dash_bootstrap_components as dbc
```

```python
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']


app = JupyterDash(__name__, external_stylesheets=external_stylesheets)


app.layout = html.Div([

    html.H1("Web Application Dashboards with Dash", style={'text-align': 'center'}),
        
    
    
    dcc.Slider(
        id='year-slider',
        min=df['review_date'].min(),
        max=df['review_date'].max(),
        value=df['review_date'].min(),
        marks={str(year): str(year) for year in df['review_date'].unique()},
        step=None
    ),
   
    dcc.Dropdown(
        id='graph-dropdown',
        
        options=[
            {'label': 'Scatter', 'value': 'scatter'},
            {'label': 'Box Plot', 'value': 'boxplot'},
            {'label': 'Bar Plot', 'value': 'barplot'},
            
        ],
        value='scatter'
    ),
            
    dcc.Graph(id='graph-with-slider'),
    
    
#     html.Label('Dropdown'),
#     dcc.Dropdown(
#         options=[
#             {'label': 'New York City', 'value': 'NYC'},
#             {'label': u'Montréal', 'value': 'MTL'},
#             {'label': 'San Francisco', 'value': 'SF'}
#         ],
#         value='MTL'
#     ),

#     html.Label('Multi-Select Dropdown'),
#     dcc.Dropdown(
#         options=[
#             {'label': 'New York City', 'value': 'NYC'},
#             {'label': u'Montréal', 'value': 'MTL'},
#             {'label': 'San Francisco', 'value': 'SF'}
#         ],
#         value=['MTL', 'SF'],
#         multi=True
#     ),

#     html.Label('Radio Items'),
#     dcc.RadioItems(
#         options=[
#             {'label': 'New York City', 'value': 'NYC'},
#             {'label': u'Montréal', 'value': 'MTL'},
#             {'label': 'San Francisco', 'value': 'SF'}
#         ],
#         value='MTL'
#     ),

#     html.Label('Checkboxes'),
#     dcc.Checklist(
#         options=[
#             {'label': 'New York City', 'value': 'NYC'},
#             {'label': u'Montréal', 'value': 'MTL'},
#             {'label': 'San Francisco', 'value': 'SF'}
#         ],
#         value=['MTL', 'SF']
#     ),

#     html.Label('Text Input'),
#     dcc.Input(id='my-input', value='initial value', type='text'),

#     html.Label('Text Output'),
#     html.Div(id='my-output'),
    
    
    
#     html.Label('Slider'),
#     dcc.Slider(
#         min=0,
#         max=9,
#         marks={i: 'Label {}'.format(i) if i == 1 else str(i) for i in range(1, 6)},
#         value=5,
#     ),

    
    
])

@app.callback(
    Output('graph-with-slider', 'figure'),
    Input('year-slider', 'value'),
    Input('graph-dropdown', 'value')
)




def update_figure(selected_year,graph_type):
    filtered_df = df[df.review_date == selected_year]

    
    if graph_type=="boxplot":
        
        fig = px.box(filtered_df, x='company_location', y='rating',
                    hover_name="company",
                    color="company_location")
 
        fig.update_layout(transition_duration=500)

    elif graph_type=="barplot":
        
        fig = px.bar(filtered_df, x='company_location', y='rating',
                    color="rating", hover_name="company",)
 
        fig.update_layout(transition_duration=500)

    elif graph_type=="scatter":
        fig = px.scatter(filtered_df, x="company_location", y="rating",
                         color="rating", hover_name="company",
                         size="counts_of_ingredients",
    #                      log_x=True, 
                         size_max=55
                        )

        fig.update_layout(transition_duration=500)

    return fig

# ------------------------------------------------------------------------------
if __name__ == '__main__':
    app.run_server(mode='inline',debug=True)
```
