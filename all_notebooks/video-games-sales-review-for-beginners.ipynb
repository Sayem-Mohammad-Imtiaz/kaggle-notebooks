# Scatter Plot

Video game purchase rate of NA Eu and JP by years

```python
plt.scatter(data["Year"],data["NA_Sales"],color="green",label="NA",alpha=0.5)
plt.xlabel("Year")
plt.legend(loc="upper right")
plt.ylabel("SALES")

plt.scatter(data["Year"],data["EU_Sales"],color="blue",label="EU",alpha=0.5)
plt.xlabel("Year")
plt.legend(loc="upper right")
plt.ylabel("SALES")

plt.scatter(data["Year"],data["JP_Sales"],color="red",label="JP",alpha=0.5)
plt.xlabel("Year")
plt.legend(loc="upper right")
plt.ylabel("SALES")

plt.show()
```
