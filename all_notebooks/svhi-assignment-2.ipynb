# a) Provide the descriptives for your data;

**Hypothesis** I'm going to analyse is as follow:

Students with **previous attempts** in a course are **likely to pass** than a student taking the course first time.

Now, I'll read the student info data and read the first 10 rows to better anaylse the data. 

```python
studentInfo = pd.read_csv('/kaggle/input/student-demographics-online-education-dataoulad/studentInfo.csv')
# students with previous attempts
studentInfo.head()
```

```python
# lets check all the final results
studentInfo["final_result"].unique()
```

```python
# Create a new column to classify final results. That will help us to do descriptive analysis!
studentInfo["final_result.class"] = 1

studentInfo["final_result.class"].loc[(studentInfo["final_result"] == "Withdrawn")] = 2
studentInfo["final_result.class"].loc[(studentInfo["final_result"] == "Fail")] = 3
studentInfo["final_result.class"].loc[(studentInfo["final_result"] == "Distinction")] = 4

#and look at the dataset again
studentInfo.head()
```

```python
factorsUsedInHypothesis = studentInfo[["id_student", "num_of_prev_attempts", 'final_result.class', 'final_result']]
factorsUsedInHypothesis.head()
```

```python
factorsUsedInHypothesis.describe()
```

```python
factorsUsedInHypothesis.plot(x='id_student', y='num_of_prev_attempts', kind='hist')
```

```python
factorsUsedInHypothesis.plot(x='id_student', y='final_result.class', kind='hist')
```

# b) Examine whether there are missing values. Explain how a missing value could occur in the data, what policy would you choose to deal with it, and why.

```python
factorsUsedInHypothesis.isna().sum()
```

In our data, there are no missing values. 

If there are missing values, we can use multiple strategies to fix that. For example, if data is very large and removing some of the rows doesn't cause any problem/deficiencies in analysis, we can delete the rows. Or we can also add zero or median/average based on our scenario. 

### num_of_prev_attempts

In this case, `num_of_prev_attempts` can be empty/null and it can be becase system is not being updated yet. Also, data can have some anomaly.

**My Solution**
I'll add `0` where `num_of_prev_attempts` are missing/null. Because in my opinion, it is okay to mark that student as `num_of_prev_attempts` zero.


### final_result

If `final_result` is empty, it can have many meanings. For example, may be student final result is not updated yet. Or may be student left the school and that student record is being updated anymore. It really depends on the situation.

**My Solution**
In this case, I'll not mark student as `pass`, `fail` etc. Because it can effect our results. So, I'll delete it to not consider this student in our data analysis. 

# c) Examine whether there are outliers. Explain what do the outliers signify in your case, what policy would you choose to deal with them, and why.

```python
fig, ax = plt.subplots(figsize=(16,8))
ax.scatter(factorsUsedInHypothesis['id_student'], factorsUsedInHypothesis['num_of_prev_attempts'])
ax.set_xlabel('Student Distribution')
ax.set_ylabel('Num of prev_attempts')
plt.show()
```

As you can see, there are very few students where `pre_attempts` is 6. It can have this implication that many courses or department doesn't allow more than 5 attempts. And this data is there by mistake or something.
