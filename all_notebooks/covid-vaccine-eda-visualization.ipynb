Let deploy graphs to analyse the data

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
df.columns
```

```python
data1=df.groupby(by=['country']).sum().reset_index()
```

```python
px.bar(data1.sort_values(by=['people_fully_vaccinated_per_hundred'], ascending=False).reset_index(),x='country',y='people_fully_vaccinated_per_hundred',title='People Vaccinated by country per hundered',color='country')
```

```python
data1.head()
```

```python
data2=df.groupby(by=['vaccines','country']).sum().reset_index()
```

```python
data2
```

```python
fig = go.Figure()
px.bar(data2.sort_values(by=['people_fully_vaccinated_per_hundred'], ascending=False).reset_index(),x='vaccines',y='people_fully_vaccinated_per_hundred',title='Vaccines taken by company',color='vaccines',hover_name='country')
```

```python
df.head()
```

```python
px.line(df,y='people_vaccinated',x=df['date'],color='country', title='No of people Vaccinated by Country')
```

```python
px.line(df,y='people_fully_vaccinated',x=df['date'],color='country', title='No of people fully Vaccinated by Country')
```

```python
df.head()
```

```python
data=df[['country','total_vaccinations','people_vaccinated','people_fully_vaccinated','vaccines','iso_code']].groupby(by=['country','iso_code','vaccines']).sum().reset_index()
```

```python
data
```

```python
px.scatter(data, x='total_vaccinations',y='people_vaccinated', color='country', size='people_vaccinated')
```

```python

px.scatter_geo(data,locations='iso_code', size='people_vaccinated',color='vaccines', title='Country by People Vaccinated')

```

```python
px.scatter_geo(data,locations='iso_code', size='people_fully_vaccinated',color='vaccines', title='Country by People fully Vaccinated')
```
