```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (32, 19)
plt.rcParams["font.size"] = 18

countrys = ['Australia', 'Austria', 'Belgium', 'Canada', 'China', 'Denmark', 'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'India', 'Ireland', 'Italy', 'Japan', 'Korea', 'Latvia', 'Lithuania', 'Luxembourg', 'Mexico', 'Netherlands', 'New Zealand', 'Norway ', 'Poland', 'Portugal', 'Slovenia', 'South Africa', 'Spain', 'Sweden', 'Turkey', 'UK', 'USA']

paidworks = [0]*len(countrys)
educations = [0]*len(countrys)
households = [0]*len(countrys)
houseworks = [0]*len(countrys)
shoppings = [0]*len(countrys)
unpaids = [0]*len(countrys)
sleeps = [0]*len(countrys)
foods = [0]*len(countrys)
personalcares = [0]*len(countrys)
sports = [0]*len(countrys)
events = [0]*len(countrys)
friends = [0]*len(countrys)
tvradios = [0]*len(countrys)
others = [0]*len(countrys)

f = open("/kaggle/input/time-usage-in-oecd-countries/Time-Use-in-OECD.csv", "r")

lines = f.readlines()

for i in range(1, len(lines)):
    line = lines[i].replace("\n", "").split(",")
    
    paidworks[countrys.index(line[0])] = int(line[1])-326
    educations[countrys.index(line[0])] = int(line[2])-57
    households[countrys.index(line[0])] = int(line[3])-61
    houseworks[countrys.index(line[0])] = int(line[4])-187
    shoppings[countrys.index(line[0])] = int(line[5])-32
    unpaids[countrys.index(line[0])] = int(line[6])-98
    sleeps[countrys.index(line[0])] = int(line[7])-553
    foods[countrys.index(line[0])] = int(line[8])-133
    personalcares[countrys.index(line[0])] = int(line[9])-107
    sports[countrys.index(line[0])] = int(line[10])-42
    events[countrys.index(line[0])] = int(line[11])-42
    friends[countrys.index(line[0])] = int(line[12])-82
    tvradios[countrys.index(line[0])] = int(line[13])-148
    others[countrys.index(line[0])] = int(line[14])-154

plt.xticks(rotation=75)
plt.grid(linestyle="--")
plt.bar(countrys, paidworks, label="Paidwork")
plt.bar(countrys, educations, label="Education", bottom=paidworks)
plt.bar(countrys, households, label="Household", bottom=educations)
plt.bar(countrys, houseworks, label="Housework", bottom=households)
plt.bar(countrys, shoppings, label="Shopping", bottom=houseworks)
plt.bar(countrys, unpaids, label="Unpaid", bottom=shoppings)
plt.bar(countrys, sleeps, label="Sleep", bottom=unpaids)
plt.bar(countrys, foods, label="Food", bottom=sleeps)
plt.bar(countrys, personalcares, label="Personalcare", bottom=foods)
plt.bar(countrys, sports, label="Sport", bottom=personalcares)
plt.bar(countrys, events, label="Event", bottom=sports)
plt.bar(countrys, friends, label="Friend", bottom=events)
plt.bar(countrys, tvradios, label="TVRadio", bottom=friends)
plt.bar(countrys, others, label="Other", bottom=tvradios)
plt.legend()
plt.xlabel("Countrys")
plt.show()


    
```
