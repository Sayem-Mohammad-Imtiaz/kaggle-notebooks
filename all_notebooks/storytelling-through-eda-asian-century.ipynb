# STORY OF THE INDIAN SUBCONTINENT

**The Indian subcontinent, Consists of India, Nepal, Bangladesh, Pakistan, Bhutan And Sri Lanka.**

![wswd](https://upload.wikimedia.org/wikipedia/commons/d/dc/Indian_Subcontinent_%28orthographic_projection%29.png)
* wikipedia

```python
df=data.iloc[[12,32,29,4,5,38],:]
df
```

```python
print('The Total Population Of The Indian Subcontinent Is',df["Population"].sum())
```

```python
print('The Mean Population Density Of The Indian Subcontinent Is',df["Density"].mean())
```

```python
print('The Mean Rate Of Population Increase Of The Indian Subcontinent Is',df["YearlyChange"].mean())
```

```python
print('The Median Age Of The People In Indian Subcontinent Is',df["MedAge"].mean())
```

```python
print('The Mean Fertility Rate Of The Indian Subcontinent Is',df["FertRate"].mean())
```

```python
print('The World Population Share Of The Indian Subcontinent Is',df["WorldShare"].sum())
```

**Nearly One In Four Humans On The Planet Lives In The Indian Subcontinent**

### Population Share Of Countries Of Indian Subcontinent Amongst Themselves

```python
sns.catplot(x="Country", y="Population", kind="bar", data=df,height=5, aspect=27/9,)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,2,2])
ax.axis('equal')
ax.pie(df["Population"], labels = df["Country"],autopct='%1.2f%%')
plt.show()
```
