```python
data["Item_Fat_Content"].nunique()
```

```python
data["Item_Fat_Content"].unique()
```

### Low fat = low fat = LF all must be have the same spelling
### Regular = reg  both must be have the same spelling

```python
data.replace("Low Fat","low fat",inplace = True)
data.replace("LF","low fat",inplace = True)
data.replace("reg","Regular",inplace = True)
```

```python
data["Item_Fat_Content"].replace({"LF": "Low fat", "reg": "Regular","Low Fat":"low fat"}, inplace=True)
```

```python
data["Item_Fat_Content"].unique()
```

```python
data.Item_Fat_Content = data.Item_Fat_Content.apply(lambda x: x.lower().title().replace("Lf", "Low Fat"))
```

```python
data["Outlet_Location_Type"].unique()
```

```python
data.groupby("Item_Fat_Content")["Item_Outlet_Sales"].sum().plot(kind = "bar");
plt.xlabel("item fat content", fontsize = 15)
plt.ylabel("Total sales count", fontsize = 15)
plt.title("relation between fat content and sales", fontsize = 15);
```

### low fat items are being sold more than regular
