```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
print(df.columns)
fig = px.choropleth(df,locations='country',locationmode='country names',color='total_vaccinations_per_hundred',hover_name='country',animation_frame='date',color_continuous_scale='orrd',range_color=(0.0,100.0))
fig.update_layout(title_text="Percentage of the coutry's total polpulation that are vaccinated",title_x=0.4)
fig.show()

plotly.offline.plot(fig,filename='vaccinations.html')
```

I chose 'total_vaccinations_per_hundred', and the color bar is a simple reddish gradient.

The darker the color, the more vaccinations have been administered, which is easier to see visually.

Also, the number of people vaccinated per 100 people can be read as a percentage.

```python
df = pd.read_csv('../input/20110311-earthquake/query (1).csv',header=0)
df.index = pd.to_datetime(df['time'])
df['time'] = df.index.strftime('%Y-%m-%d %H:00:00')
df = df.sort_index()
fig = px.scatter_geo(df,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='rainbow',range_color=(1.0,10))
fig.show()
```

I chose the Tohoku Earthquake that occurred at 14:46 (Japan time) on March 11, 2011.

From the start time, we can see that there were many aftershocks along the east coast of Japan.

There have been more than 14,000 aftershocks so far, but on April 1 this year, the Japan Meteorological Agency (JMA) dropped the method of announcing that an earthquake with a seismic intensity of lower-5 or higher in the aftershock area was "considered an aftershock.

This is because after 10 years, it is no longer possible to clearly determine that an earthquake is an aftershock, and also because earthquakes of magnitude 7 to 8 have been occurring repeatedly along the Japan Trench for a long time.
