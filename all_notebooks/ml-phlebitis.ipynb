```python
#Get the dependencies
import pandas as pd
import numpy as np
from xgboost import XGBClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
#Load the datasets into data frame
df = pd.read_csv('../input/phlebitis-score-temperature/3. Validation_filtered_Duplicate.csv')
df.head()
```

```python
#Check datasets for any missing values
df.isnull().values.any()
```

```python
#Get the number rows and columns in the datasets
df.shape
```

```python
#Get target count
#Identify which patients has phlebitis and which patients don't have phlebitis
df['VipGroup'].value_counts()
```

```python
#Guess the probability of Phlebitis Occurrence
percent_with_phebitis = 55/(55+82)*100
percent_without_phlebitis = 82/(55+82)*100
print('If I guess the individual did not have Phlebitis, I would be correct ',percent_without_phlebitis,'% of the time' )
print('If I guess the individual has Phlebitis, I would be correct ',percent_with_phebitis,'% of the time' )
```

```python
#Visualize the count
sns.countplot(df['VipGroup'])
```

```python
#Get the data types
df.dtypes
```

```python
#Create the feature datasets
X = df.drop(['AnonymizedID'], 1)
X = np.array(X.drop(['VipGroup'], 1))
#Create the target datasets
y = np.array(df['VipGroup'])
```

```python
#Split the data into 80% training and 20% testing datasets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
#Transform feature data to be values between 0 and 1
Scale = MinMaxScaler(feature_range = (0,1))
x_train = Scale.fit_transform(x_train)
x_test = Scale.transform(x_test)
```

```python
#Create XGBClassifier (Extreme Grdient boosting model based on the decision tree)
#XGBoost is the new ML algorithm designed with speed & performance
#train model on x_train & y_train
Model = XGBClassifier().fit(x_train, y_train)
```

```python
#Get the model predictions
#Will predict the y_test data based on x_test data
predictions = Model.predict(x_test)
predictions
```

```python
y_test
```

```python
#Evaluate the model. Get model accuracy, precision, recall, and f1-score
print(classification_report(y_test, predictions))
```
