```python
data = promo.sample(frac=0.8, random_state=42)
data_unseen = promo.drop(data.index)
data.reset_index(inplace=True, drop=True)
data_unseen.reset_index(inplace=True, drop=True)
print('Data for Modeling: ' + str(data.shape))
```

```python
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.classification import *
promo= setup(data =data, session_id=42,
             target = 'Target',
                  normalize = True, 
                  transformation = True, 
                  log_experiment = True,
                  remove_multicollinearity = True, #rop one of the two features that are highly correlated with each other
                  date_features=['RegisteredOn','Dob'],
            categorical_features=[ 'Gender', 'UserId', 'Mobile', 'PostalCode',
       'OS Type', 'Reg Referral Code', 'Reg Referral Prefix',
       'Ref PC AC Number', 'Name', 'Sim_OS', 'Sim_Info', 'Sim_User',
       'Device_OS', 'Device_Info', 'Device_User', 'LastLoginTime_HH',
       'LastLoginTime_MM', 'LastLoginTime_SS', 'PresentLoginTime_HH',
       'PresentLoginTime_MM', 'PresentLoginTime_SS', 'Meta Data_2',
       'Meta Data_4', 'Email_user', 'Reg Client IP_1', 'Reg Client IP_2',
       'Reg Client IP_3', 'Reg Client IP_4'],
            high_cardinality_features =[ 'UserId', 'Mobile', 'PostalCode',
       'OS Type', 'Reg Referral Code',
       'Ref PC AC Number', 'Name', 'Sim_User',
        'Device_User', 'LastLoginTime_HH',
       'LastLoginTime_MM', 'LastLoginTime_SS', 'PresentLoginTime_HH',
       'PresentLoginTime_MM', 'PresentLoginTime_SS',
       'Meta Data_4', 'Email_user', 'Reg Client IP_1', 'Reg Client IP_2',
       'Reg Client IP_3', 'Reg Client IP_4']  ,         
            train_size = 0.8)
```

```python
lr = create_model('lr')
```

```python
print(lr)
```

```python
tuned_lr = tune_model(lr,optimize = 'F1')
```

```python
plot_model(tuned_lr, plot = 'auc')
```

```python
plot_model(tuned_lr, plot = 'pr')
```

```python
plot_model(tuned_lr, plot='feature')
```

```python
plot_model(tuned_lr, plot = 'confusion_matrix')
```

```python
evaluate_model(tuned_lr)
```

```python
predict_model(tuned_lr)
```

```python
unseen_predictions = predict_model(tuned_lr, data=data_unseen)
unseen_predictions.head()
```

```python
print("Confidence Score :   {}".format(round(unseen_predictions.Score.mean(),2)))#Confidence Score
```
