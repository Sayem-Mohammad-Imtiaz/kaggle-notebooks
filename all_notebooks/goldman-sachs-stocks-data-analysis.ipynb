```python
f, ax = plt.subplots(2, 2, figsize=(15,10), sharex=True)
f.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
f.gca().xaxis.set_major_locator(mdates.YearLocator())

ax[0,0].plot(stocks.index, stocks['Open'], color='r')
ax[0,0].grid(True)
ax[0,0].tick_params(labelrotation=90)
ax[0,0].set_title('Open Price');

ax[0,1].plot(stocks.index, stocks['Close'], color='g')
ax[0,1].grid(True)
ax[0,1].tick_params(labelrotation=90)
ax[0,1].set_title('Close Price');

ax[1,0].plot(stocks.index, stocks['High'], color='b')
ax[1,0].grid(True)
ax[1,0].tick_params(labelrotation=90)
ax[1,0].set_title('Highest Price');

ax[1,1].plot(stocks.index, stocks['Low'], color='y')
ax[1,1].grid(True)
ax[1,1].tick_params(labelrotation=90)
ax[1,1].set_title('Lowest Price');
```

```python
plt.figure(figsize=(14,5))
plt.plot(stocks.index, stocks['Volume'],color='purple')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())
plt.xticks(rotation=90)
plt.title('Number of units traded in a day')
plt.show()
```

### > 2008 had highest number of units traded in a day. 

## **Which month of year 2008 had most number of units traded?**

```python
plt.figure(figsize=(14,5))
df = stocks.loc[pd.Timestamp('2008-01-01'):pd.Timestamp('2008-12-31')]
plt.scatter(df.index, df['Volume'],color='red')
plt.grid(True)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=90)
plt.show()
```

### > September month of year 2008 had most number of units traded. 

## **Which day was it?**

```python
plt.figure(figsize=(14,5))
df = stocks.loc[pd.Timestamp('2008-09-01'):pd.Timestamp('2008-09-30')]
plt.plot(df.index, df['Volume'],color='green')
plt.grid(True)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator())
plt.xticks(rotation=90)
plt.fill_between(df.index, df['Volume'],color='green')
plt.show()
```

### > On 18 September 2008, highest number of units were traded.  

## **Adjusted Close price for the year 2021**

```python
stock21 = stocks.loc[pd.Timestamp('2021-01-01'):pd.Timestamp('2021-07-22')]
```

```python
weekly_stock = stock21.resample('W').mean() # resampling weekly
w1=weekly_stock.head()
w1.style.background_gradient()
```

```python
plt.figure(figsize=(12,5))
plt.plot(weekly_stock.index, weekly_stock['Adj Close'], '-*',color='yellow')
plt.grid(True)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45)
plt.show()
```

## **Analysing difference between open and close prices**

```python
stocks['diff'] = stocks['Open'] - stocks['Close']
stocks_diff = stocks.resample('W').mean() # resampling weekly
sd=stocks_diff.tail(10)
# sd.style.background_gradient().highlight_min(axis=0, color='red').highlight_max(axis=0, color='green')
def color_positive_green(val):
    if val > 0:
        color = 'green'
    else:
        color = 'red'
    return 'color: %s' % color
 
sd.style.applymap(color_positive_green)
```
