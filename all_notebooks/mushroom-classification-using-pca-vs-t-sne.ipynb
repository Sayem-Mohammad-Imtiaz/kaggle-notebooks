```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('../input/mushroom-classification/mushrooms.csv')
df.head()
```

```python
X=df.drop('class', 1)
y=df['class']
```

```python
y = y.map({'p': 'Posionous', 'e': 'Edible'})
```

```python
X.info()
```

* Since all the input features are object types. we need to encode it into numeric
* Let's check the number of unique values in each feature

```python
cat_cols= X.select_dtypes(include='object').columns.tolist()
for col in cat_cols:
    print (f" col name : {col}, N Unique : {X[col].nunique()}")
```

* Since the number of classes in each features are not very high, we can chage the features into categories and encode them.
* We need not to worry about the priorities in categorical encoding using the following method.

```python
for col in cat_cols:
    X[col]=X[col].astype("category")
    X[col]=X[col].cat.codes
```

```python
X.head()
```

```python
from sklearn.preprocessing import StandardScaler
X_std= StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA
```

```python
X_pca = PCA(n_components=2).fit_transform(X_std)
X_pca
```

```python
X_pca = np.vstack((X_pca.T, y)).T
X_pca
```

```python
df_pca = pd.DataFrame(X_pca, columns=['1st_Component', '2nd_Component', 'class'])
df_pca.head()
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(data=df_pca, hue='class', x='1st_Component', y='2nd_Component')
plt.show()
```

```python
from sklearn.manifold import TSNE

X_tsne = TSNE().fit_transform(X_std)
X_tsne = np.vstack((X_tsne.T, y)).T
df_tsne = pd.DataFrame(X_tsne, columns=['Dim1', 'Dim2', 'class'])
df_tsne.head()
```

```python
plt.figure(figsize=(8, 8))
sns.scatterplot(data=df_tsne, hue='class', x='Dim1', y='Dim2')
plt.show()
```

by this resule we can see that t-SNE outperforms than PCA
