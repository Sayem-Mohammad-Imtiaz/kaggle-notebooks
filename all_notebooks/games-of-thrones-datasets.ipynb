## Correlations ##
Let's take a look at how everything is correlated within their datasets, and how they are correlated to each other. A positive number indicates that as x increases, so does y. A negative number indicates that as x increases, y decreases. The closer to 1 or -1 the number is, the more relationship they have between them. Therefore, the closer to 0 a number is, there is less correlation between the two.

```python
battles.corr()
```

```python
deaths.corr()
```

```python
predictions.corr()
```

Let's now look at how things are correlated between the 2 datasets. Is there any correlation between the year of a battle and the gender?

```python
battles['year'].corr(deaths['Gender'])
```

As we can see from above, there is no correlation between the 2 bits of information.

## Correlations in Heatmaps ##
Let's now create a heatmap so that we can visualize the correlation between different bits of information.

```python
correlation = battles.corr()
mp.figure(figsize = (10,10))
sn.heatmap(correlation, vmax=1, square=True, annot=True,cmap='cubehelix')

mp.title("Attacking and Defending Correlation")
```

From the above information, we can draw several conclusions.

 1. There is a strong correlation between the year the battle was found and the number of battles
 2. There is a very high correlation between the fact that as the years go on, there are less summer battles.

```python
correlation = deaths.corr()
mp.figure(figsize = (10,10))
sn.heatmap(correlation, vmax=1, square=True, annot=True,cmap='cubehelix')

mp.title("Character Deaths")
```

```python
correlation = predictions.corr()
mp.figure(figsize = (15,15))
sn.heatmap(correlation, vmax=1, square=True, annot=True,cmap='cubehelix')

mp.title("Predictions of Death")
```

## Kings and Battles ##

Now looks look at how many times each king has confronted each other in battle.

```python
battles.groupby(['attacker_king', 'defender_king']).count()['name'].plot(kind = 'barh')
```

Now that we know how many times each king has been in battle with each other, how many times has each king won/loss?

```python
battles.groupby(['attacker_king', 'attacker_outcome']).count()['name'].unstack().plot(kind = 'barh')
```

Now let's look at which king was involved in the most attacks using wordcloud.

```python
attackers = battles.attacker_king.map(lambda x:str(x).split(","))
empty_array = []
for i in attackers:
    empty_array = np.append(empty_array, i)
```

```python
from wordcloud import WordCloud
cloud = WordCloud(width=1440, height=1080, relative_scaling=0.5, stopwords=['battle']).generate(" ".join(empty_array))
mp.figure(figsize=(20, 15))
mp.imshow(cloud)
mp.axis('off')
mp.show()
```

As we can see from the above, Joffrey Tommen and Tommen Baratheon were mentioned in more battles than attackers like Euron Greyjoy.
