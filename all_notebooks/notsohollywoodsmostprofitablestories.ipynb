<h2><span style = "font-family: fantasy;background-color:#D2AD62;">Columns Description</span></h2>


- Film : Name of film
- Genre : Genre of the movie 
- Lead Studio : Name of the lead studio 
- Audience Score :  The Audience Score, denoted by a popcorn bucket, is the percentage of users who have rated the movie or TV Show positively. For films to which we can verify users have bought tickets, the Audience Score is made up of Verified Ratings
- Profitability : % of profit (need to verify) 
- Rotten Tomatoes % : The percentage is based on the film's reviews aggregated by the website and assessed as positive or negative, and when all aggregated reviews are positive, the film has a 100% rating
- Worldwide Gross : Number of tickets sold x ticket price. It's the total amount generated by a movie at. ticket counters
- Year : Year of release

<h1><span style = "font-family: fantasy;background-color:#D2AD62;">Checking: missing values</span></h1>

```python
missingno.matrix(data, sparkline=False, figsize=(10,5), fontsize=12, color=(0.27, 0.52, 1.0));
```

```python
missingno.heatmap(data, figsize=(10,5), fontsize=12);
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['axes.facecolor'] = 'gray'
```

```python
fig = plt.figure(figsize=(15,7))

ax1 = fig.add_subplot(1,2,1)
missingno.dendrogram(data, orientation="right", method="centroid", fontsize=12, ax=ax1);

ax2 = fig.add_subplot(1,2,2)
missingno.dendrogram(data, orientation="top", method="ward", fontsize=12, ax=ax2);

plt.tight_layout()
```
