```python
Iris.drop(["Id","SepalLengthCm"], axis =1)
```

```python
x= np.array(Iris.loc[:,"PetalWidthCm"]).reshape(-1,1)
y= np.array(Iris.loc[:,"Species"]).reshape(-1,1)


```

```python
print(Iris.Species.unique())
```

```python
x.shape
```

```python
#Plotting 
plt.figure(figsize=[10,10])
plt.scatter(x=x,y=y)
plt.xlabel('PetalWidthCm')
plt.ylabel('Species')
plt.show()
```

```python
#Linear Regression 
from sklearn.linear_model import LinearRegression
reg = LinearRegression()

#Change y the output the categorical to numeric values
Iris["Species"]= [0 if i =="Iris-setosa" else 1 if i == "Iris-versicolor" else 2 i == "Iris-virginica" for i in Iris["Species"]]

# Predict space
predict_space = np.linspace(min(x), max(x)).reshape(-1,1)

# Fit
reg.fit(x,y)

# Predict
predicted = reg.predict(predict_space)

# R^2 
print('R^2 score: ',reg.score(x, y))

# Plot regression line and scatter
plt.plot(predict_space, predicted, color='black', linewidth=3)
plt.scatter(x=x,y=y)
plt.xlabel('PetalWidthCm')
plt.ylabel('Species')
plt.show()
```
