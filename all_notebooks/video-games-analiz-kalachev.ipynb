# **Игры, у которых оценка выше 5 по статистике Reddit**

```python
richgames = df.loc[df['Score'] >= 5]
richgames.head()
```

```python
plt.scatter(df['Score'].iloc[:20],df['GameName'].iloc[:20],color='r')
```

На графике видно, что большинство популярных игр оценивают в среднем 7 баллов.

# **Игры, у которых оценка 5 или ниже по статистике Reddit**

```python
poorgames = df.loc[df['Score'] <= 5]
poorgames.head()
```

Как мы видим, игр у которых оценка ниже 5 меньше чем больше 5, по анализу рынка мы можем сделать вывод, что в среднем игрок доволен игрой, которую купил.

# **Диаграмма игр и их оценка на 2021 год**

```python
fig = plt.figure(figsize=(9,50))
fig = plt.plot(df['Review'].iloc[:200],df['GameName'].iloc[:200])
```

Как мы можем заметить, есть игры с хорошей оценкой, так и с плохой, но всё так же диграмма показыват что в среднем игрок доволен.

# **Диаграмма отзывов о играх на 2021 год **

```python
df.Review.unique()
plt.pie(df['Review'].value_counts(),labels=['Хороший', 'Ранний доступ', 'Отличный', 'Превосходный', 'Справедливый', 'Ужасный',
       'Посредственный', 'Плохой', 'Бедный', 'Важный', 'Бездонный'])
```

Как мы можем заметить, в среднем об играх отзываются "Хорошо" и "Отлично",а так же множество игр находятся на "Раннем доступе", то есть нуждаются в доработке
