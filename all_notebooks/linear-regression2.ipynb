```python
import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv("../input/startup-logistic-regression/50_Startups.csv")
print(df.info())
```

```python
states = df.State.unique() # print out unique states
states
```

```python
# Using filter seperate data state by state
df_newyork = df[df.State == "New York"]
df_california = df[df.State == "California"]
df_florida = df[df.State == "Florida"]
```

```python
# For New York
x1 = df_newyork["R&D Spend"].values.reshape(-1, 1)
x2 = df_newyork["Administration"].values.reshape(-1, 1)
x3 = df_newyork["Marketing Spend"].values.reshape(-1, 1)
y = df_newyork["Profit"].values.reshape(-1, 1)

# To train data, I use 80 percent of the data as usual
y_train = y[0:int(len(y)*8/10)]
x1_train = x1[0:int(len(x1)*8/10)]
x2_train = x2[0:int(len(x2)*8/10)]
x3_train = x3[0:int(len(x3)*8/10)]
```

```python
linear_regression = LinearRegression()
model_x1 = linear_regression.fit(x1_train, y_train) # train model
y_head_x1 = model_x1.predict(x1) # test
print("R^2 Score of Regression Analysis of R&D Spend and Profit for New York = ", r2_score(y, y_head_x1))
```

```python
plt.figure(figsize = (10, 8))
plt.scatter(x1, y, color = "blue", label = "Real Data")
plt.plot(x1, y_head_x1, color = "red", label = "Regression Line")
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.title("New York")
plt.legend()
plt.show()
```

```python
model_x2 = linear_regression.fit(x2_train, y_train) # train model
y_head_x2 = model_x2.predict(x2) # test
print("R^2 Score of Regression Analysis of Administration and Profit for New York = ", r2_score(y, y_head_x2))
```

```python
plt.figure(figsize = (10, 8))
plt.scatter(x2, y, color = "blue", label = "Real Data")
plt.plot(x2, y_head_x2, color = "red", label = "Regression Line")
plt.xlabel("Administration")
plt.ylabel("Profit")
plt.title("New York")
plt.legend()
plt.show()
```

```python
model_x3 = linear_regression.fit(x3_train, y_train) # train model
y_head_x3 = model_x3.predict(x3) # test
print("R^2 Score of Regression Analysis of Marketing Spend and Profit for New York = ", r2_score(y, y_head_x3))
```
