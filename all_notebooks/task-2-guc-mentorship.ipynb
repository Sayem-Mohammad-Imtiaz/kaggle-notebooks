```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
```

**Regression Data SET:**

```python
train_reg= pd.read_csv('../input/heart-disease-uci/heart.csv')
train_reg.head()

```

Horray! All Data are Numerical!

Checking for Nulls:

```python
train_reg.info()
```

```python
##Separating Target from Other Features :
X=train_reg.drop('target', axis=1)
y= train_reg['target']
##Splitting Data set to train and test:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Scaling dataset to be used with linear regression and SVR:



```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Sc_X_train=scaler.fit_transform(X_train)
Sc_X_test=scaler.transform(X_test) 
```

```python
lin_reg = LinearRegression(normalize=True)
lin_reg.fit(Sc_X_train,y_train)
```

```python
score_train = lin_reg.score(Sc_X_train, y_train) #Coefficient of determination
score_test =lin_reg.score(Sc_X_test, y_test)

print('train score: {}'.format(score_train)) 
print('test score: {}'.format(score_test))
```

Too bad? :(

Predicting...

```python
y_pred =lin_reg.predict(Sc_X_test)
y_pred
```

```python
compare = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
compare
```

Since the target consists only of 0s or 1s , this is an attempt to round the predicted outputs

```python
compare = pd.DataFrame({'Actual': y_test, 'Predicted': np.fromiter((round(xi) for xi in y_pred), y_pred.dtype)})
compare
```

```python
from sklearn.svm import SVR

reg=SVR(kernel='rbf')
reg.fit(X_train, y_train)


svr_score_train = reg.score(X_train, y_train) #Coefficient of determination
svr_score_test =reg.score(X_test, y_test)

print('train score: {}'.format(svr_score_train)) 
print('test score: {}'.format(svr_score_test))
```

Classification dataset

```python
train_C = pd.read_csv('../input/heart-disease-uci/heart.csv')
train_C.info()
```

```python
##Separating Target from Other Features :
X=train_C.drop('target', axis=1)
y= train_C['target']
##Splitting Data set to train and test:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

**Decision Tree**

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier(max_depth=5)

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

**Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier

#Create a Gaussian Classifier
clf=RandomForestClassifier(n_estimators=100)

#Train the model using the training sets y_pred=clf.predict(X_test)
clf.fit(X_train,y_train)

y_pred=clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
