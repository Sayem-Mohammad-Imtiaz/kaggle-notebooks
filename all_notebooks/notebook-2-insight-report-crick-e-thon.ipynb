```python
total_run_of_players = deliveri_data.groupby(['batsman'])['total_runs'].count()
total_run_of_players = pd.DataFrame(total_run_of_players)
top_20_runners = total_run_of_players.sort_values(by='total_runs',ascending=False).head(20)
```

```python
top_20_runners.plot(kind='bar',
                   title = 'Top 20 Runners',
                   color=sns.color_palette('BuPu_r',100),
                   figsize=[20,10])
plt.show()
```

```python
batsman_with_most_number_of_boundires = deliveri_data.groupby('batsman')['batsman_runs'].agg(lambda x : 
                    (x == 4).sum()).reset_index().sort_values(by='batsman_runs',
                    ascending=False).reset_index(drop=True)
top_20_players_with_most_boundires = batsman_with_most_number_of_boundires[:20]
```

```python
boundires = deliveri_data.groupby('batsman')['batsman_runs'].agg(lambda x : 
                    (x == 4).sum())
boundires = pd.DataFrame(boundires)
top_20_boundires_played = boundires.sort_values(by='batsman_runs',ascending=False).head(20)
top_20_boundires_played.plot(kind='bar',
                   title = 'Top 20 Played Boundries by Players',
                   
                   figsize=[20,10])
plt.show()
```

```python
batsman_with_most_number_of_boundires = deliveri_data.groupby('batsman')['batsman_runs'].agg(lambda x : 
                    (x == 6).sum()).reset_index().sort_values(by='batsman_runs',
                    ascending=False).reset_index(drop=True)
top_20_players_with_most_boundires = batsman_with_most_number_of_boundires[:20]
boundires = deliveri_data.groupby('batsman')['batsman_runs'].agg(lambda x : 
                    (x == 6).sum())
boundires = pd.DataFrame(boundires)
top_20_boundires_played = boundires.sort_values(by='batsman_runs',ascending=False).head(20)
top_20_boundires_played.plot(kind='bar',
                   title = 'most sixer',
                   color=sns.color_palette('autumn',2),
                   figsize=[20,10])
plt.show()
```
