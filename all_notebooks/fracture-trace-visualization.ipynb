```python
# File handling
from pathlib import Path
# Ignore geopandas warnings
import warnings

# geopandas for spatial data management
import geopandas as gpd
# geopandas uses shapely for individual geometry abstraction
from shapely.geometry import LineString
```

```python
# Get paths of all traces (filename contains "_traces")
all_geopackages = list(Path("../input/getaberget-fracture-trace-dataset/ahvenanmaa").rglob("*_traces.gpkg"))
len(all_geopackages)
```

```python
# Filter geopandas read file warning
warnings.filterwarnings("ignore")

# Iterate over the geopackage paths
for geopackage in all_geopackages:
    
    # Read the GeoPackage file into a GeoDataFrame with geopandas
    gdf = gpd.read_file(geopackage)
    
    # Geometry type for all geometries in GeoDataFrame is LineString
    assert all([isinstance(geom, LineString) for geom in gdf.geometry.values])
    
    # Plot the GeoDataFrame
    gdf.plot(color="black", linewidth=0.5)
    
```
