1) Erstellen Sie einen Boxplot und ein Histogramm der Daten

```python
#Anzeigen von Daten aus der sun.csv-Datei
df = pd.read_csv('../input/sonnenstunden-in-thuringen-19512020/sun.csv')
df.head()
```

```python
#Erstellung einen Boxplot
plt.figure(figsize = (20, 12))
ax = sns.boxplot(x='Jahr', y='Stunden', data=df)
plt.setp(ax.artists, alpha=.5, linewidth=2, edgecolor="k")
plt.xticks(rotation=45)
plt.title('Boxplot Anzahl an Sonnenstunden in Thuringen')
```

```python
#Erstellung einen Boxplot
plt.figure(figsize = (20, 12))
ax = sns.boxplot(x='Stunden', data=df)
plt.setp(ax.artists, alpha=.5, linewidth=2, edgecolor="k")
plt.xticks(rotation=45)
plt.title('Boxplot Anzahl an Sonnenstunden in Thuringen')
```

```python
#Histogramm
# An "interface" to matplotlib.axes.Axes.hist() method
n, bins, patches = plt.hist(x=df, bins='auto', alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.title('Histogramm')
```

3) Bestimmen Sie den Median sowie das obere / untere Quartil und den Interquartilsabstand der Daten

```python
stats=pd.DataFrame()
stats["median"]=df.median()
```

```python
stats.T
```

```python
# Computing IQR
Q1 = df['Stunden'].quantile(0.25)
Q3 = df['Stunden'].quantile(0.75)
IQR = Q3 - Q1

# Filtering Values between Q1-1.5IQR and Q3+1.5IQR
filtered = df.query('(@Q1 - 1.5 * @IQR) <= Stunden <= (@Q3 + 1.5 * @IQR)')
```

```python
df.join(filtered, rsuffix='_filtered').boxplot()
```
