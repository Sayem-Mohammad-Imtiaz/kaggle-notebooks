```python
#%%

# Forced directed graph representing Hindu gods and their names
# Source: https://github.com/ak0029/Names/blob/main/Names.csv

#%%

# Import d3fdgraph and pandas

#%%

import d3fdgraph as d3fd;
import pandas as pd;

#%%

# Read Data

#%%

names = pd.read_csv("../input/names-df/Names.csv").to_dict('records');

#%%

# Function to transform data to d3fdgraph format

#%%

def getNames():
    list = []
    lastName = ""
    counter = 0;
    for row in names:
        
        if (lastName == row["belongs_to"]):
            counter += 1
        else:
            counter = 1;
            lastName = row["belongs_to"]
            
        list.append({
            "source": row["belongs_to"],
            "target": row["name"],
            "weight": 1
        })
        
    return pd.DataFrame(list)

#%%

graphData = getNames()

#%%

# Print graph data

#%%

graphData

#%%

# Display forced directed graph

#%%

d3fd.plot_force_directed_graph(graphData, node_radius=8, link_distance=100, collision_scale=4)

#%%
```
