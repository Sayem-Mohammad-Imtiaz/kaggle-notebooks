Function Used:-
1.value_counts()- Return a Series containing counts of unique values.
2.groupby()-Function is used to split the data into groups based on some criteria.
3.rename()-Method is used to rename any index, column or row.
4.duplicated()-Method helps in analyzing duplicate values only.
5.pivot_table()-Function takes simple data frame as input, and performs grouped operations.
6.sort_value()-Sort a series in ascending order by default.
7.rank()-Method returns a rank of every respective index of a series passed.
8.max()-Return a max value of a series.
9.min()-Return a min value of a series.
10.apply()-Allow the users to pass a function and apply it on every single value of the Pandas series.
11.head() - Returns the first five elements of the dataset by default.
12.tail() - Returns the last five elements of the dataset by default.
13.count() - Returns the number of occurrences of a substring in the given string
