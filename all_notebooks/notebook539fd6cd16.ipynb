# Improved visualization

From here, I will be using the modules ipot and plotly to make better visualizations.

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_new_cases = go.Scatter(x=df1.index,y=df1['daily_new_cases'],yaxis='y1',name='Daily confirmed cases')
daily_vaccinations = go.Scatter(x=df2.index,y=df2['daily_vaccinations'],yaxis='y2',name='Daily vaccinations')

layout_obj = go.Layout(title='COVID vs. Vaccines in Germany',xaxis=dict(title='Dete'),yaxis=dict(title='Daily cases'),yaxis2=dict(title='Vaccinations',side='right',overlaying='y'))
fig = go.Figure(data=[daily_new_cases,daily_vaccinations],layout=layout_obj)
iplot(fig)
fig.write_html('outcome.html')
```

# Create a global map for COVID-19 (Percentage of the population of the population that they are fully vaccinated).

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
print(df.columns)
fig = px.choropleth(df,locations='country',locationmode='country names',color='people_fully_vaccinated_per_hundred',hover_name='country',animation_frame='date',color_continuous_scale='rainbow',range_color=(0.0,20.0))
fig.update_layout(title_text="Percentage of the country's total populations that are fully vaccinated",title_x=0.5)
fig.show()

plotly.offline.plot(fig,filename='vaccinations.html')
```

# explanation of report No.7 - 1
# what I’m visualizing
・This animation show the "Percentage of the country's total populations that are fully vaccinated" between Decenber 2020 and May 2021.However,the range of values is from 0% to 20%.This range is based on vaccination rate of Germany.
# why I selected it.
・In Report No. 6, I discussed the actual status of vaccination in Germany, where we know that as of mid-May, the vaccination rate was about 17.44%.
The distribution around December 2020 also shows that Germany was the first country in the world to receive the vaccine worldwide.
In other words, by comparing the speed of vaccination progress in other countries with the progress in Germany, where vaccination first started, we can get a rough idea of the pace (fast or slow) at which vaccines are being administered in each country.
