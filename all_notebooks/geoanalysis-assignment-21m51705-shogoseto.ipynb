```python
df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1 = df1.groupby('country').sum()['daily_new_cases'].sort_values(ascending=False)
print(df1.to_string())
```

```python
selected_country_1 = 'Germany'
selected_country_2 = 'Germany'

# header=0 means that the first row is zero
df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
# filter out the country
df1 = df1[df1['country']==selected_country_1]
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2 = df2[df2['country']==selected_country_2]
# Recognize the index as a 'date' to make xaxis time
df1.index = pd.to_datetime(df1['date'])
df2.index = pd.to_datetime(df2['date'])
```

```python
print(df1.columns)
print(df2.columns)
# ここでINDEXの名前を確認
```

# Analysis

From here, I will analyze the data that was loaded as an dataframe object.

```python
# time-series data of the selected country
fig,ax = plt.subplots(figsize=(10,5))# two times larger than before
ax.plot(df1.index,df1['daily_new_cases'],label='Cases')

ax1 = ax.twinx()# areate a twin of axis
ax1.plot(df2.index,df2['daily_vaccinations'],'r',label='Vaccinations')# 'r' means Red
ax.legend()#create a legend(凡例)
ax1.legend()
plt.show()
```

# 2-1. Construct a time-series of the following: 
## ①Daily COVID-19 cases
## ②Daily vaccinations

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_new_cases = go.Scatter(x=df1.index,y=df1['daily_new_cases'],yaxis='y1',name='Daily confirmed cases')
daily_vaccinations = go.Scatter(x=df2.index,y=df2['daily_vaccinations'],yaxis='y2',name='Daily vaccinations')

#insert 'Layout' object to combine all of this
layout_obj = go.Layout(title='COVID vs. Vaccines',xaxis=dict(title='Date'),yaxis=dict(title='Daily cases'),yaxis2=dict(title='Vaccinations',side='right',overlaying='y'))
fig = go.Figure(data=[daily_new_cases,daily_vaccinations],layout=layout_obj)
iplot(fig)
fig.write_html('outcome.html')
```

# 3. What is the global ranking of that country in terms of the total number of cases
# and the total vaccinations per number of people?

```python
df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1 = df1.groupby('country').sum()['daily_new_cases'].sort_values(ascending=False)#国ごとに集計→新規感染者のデータのみ表示→降順に並べ替え(ascending=Trueだと昇順)→Top10を表示
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2 = df2.groupby('country').max()['people_fully_vaccinated_per_hundred'].sort_values(ascending=False)
#Know the ranking by getting the number of the sorted row.
rank1 = df1.index.get_loc(selected_country_1)+1
rank2 = df2.index.get_loc(selected_country_2)+1

print('Answer: The global ranking of the total number of cases in '+selected_country_1+' is '+str(rank1))#https://note.nkmk.me/python-pandas-get-loc-row-column-num/
print('Answer: The global ranking of the total vaccinations per number in '+selected_country_1+' is '+str(rank2))
```
