```python
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
plt.style.use('dark_background')
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

import warnings
warnings.simplefilter(action='ignore', category=Warning)
```

```python
df = pd.read_csv('../input/greenhouse-gas-emissions/GHG-Emissions-by-sector.csv')
```

```python
df.describe().T.style.bar(
    subset=['mean'],
    color='#65db7f').background_gradient(
    subset=['std'], cmap='magma').background_gradient(subset=['50%'], cmap='magma')
```

```python
sorted_df = df.sort_values('GlobalShare', ascending=True).reset_index()
```

```python
#T shape Plot
fig = px.bar(sorted_df, x = 'GlobalShare',color="Sub-sector", template = 'plotly_dark', title = 'Distrbution of Global Share')
fig.show()
```

```python
#histogram
fig = px.histogram(sorted_df,
                   x="GlobalShare" ,marginal="rug",color= 'Sub-sector',
                   hover_data=df.columns, template = 'plotly_dark', title = 'Distribution of GlobalShare represented by SubSector')
fig.show()
```

```python
fig = px.scatter(df,
                 x="GlobalShare", y=df.index, color= 'Sub-sector',
                 log_x=True, size_max=60,
                 template='plotly_dark', title='Scatter plot for Global Share')
fig.show()
```

```python
fig = px.pie(df, names='Sub-sector', values='GlobalShare', template = 'plotly_dark', 
             title = 'Representation of Sub-Sectors with their Global Shares')
fig.show()
```

```python
fig = px.box(df, y="GlobalShare" ,template = 'plotly_dark', 
             title = 'Representation of Global Shares')
fig.show()
```

```python
#violin plots
fig = px.violin(df, y="GlobalShare",template = 'plotly_dark')
fig.show()
```

```python
#parallel plots
fig = px.parallel_categories(df,template = 'plotly_dark', color = 'GlobalShare')

fig.show()
```
