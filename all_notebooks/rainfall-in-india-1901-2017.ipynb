```python
import numpy as np
import pandas as pd
import os
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
data0=pd.read_csv('../input/rainfall-data-from-1901-to-2017-for-india/Rainfall_Data_LL.csv')
data0
```

```python
data0['SUBDIVISION'].value_counts()
```

```python
index0=data0[data0['SUBDIVISION']=='Andaman & Nicobar Islands'].index.tolist()
index1=data0[data0['SUBDIVISION']=='Lakshadweep'].index.tolist()
index2=data0[data0['SUBDIVISION']=='Arunachal Pradesh'].index.tolist()
index3=index0+index1+index2
```

```python
print(data0.columns.tolist())
```

```python
data1=data0.drop(index=data0.index[index3])
data2=data1[['YEAR','ANNUAL']]
data3=data2.groupby('YEAR',as_index=False).mean()
data3
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=data3['YEAR'],y=data3['ANNUAL'],name="Yearly Annual Rainfall"),secondary_y=False,)
fig.add_trace(go.Scatter(x=data3['YEAR'],y=data3['ANNUAL'].rolling(window=10).mean(),name="10-Year Rollig Annual Rainfall"),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="Annual Rainfall in India (1901-2017)")
fig.update_xaxes(title_text="Year")
fig.update_yaxes(title_text="Rainfall",secondary_y=False)
fig.show()
```
