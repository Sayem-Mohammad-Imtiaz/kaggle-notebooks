```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math as math
```

```python
data=pd.read_csv('../input/nagraj/iris.data',header=None,names=['sepal_length','sepal_width','petal_length','petal_width','class'])
```

```python
 #Feature(Sepal Length) Vs Feature(Sepal Width)
plt.scatter(data.iloc[0:50,1],data.iloc[0:50,0],c='r')
plt.scatter(data.iloc[50:100,1],data.iloc[50:100,0],c='b')
plt.scatter(data.iloc[100:150,1],data.iloc[100:150,0],c='y')
plt.xlabel("Sepal Width")
plt.ylabel("Sepal Length")
plt.title("Sepal Length Vs Sepal Width")
plt.legend(labels=['setosa','versicolor','virginica'])
plt.show()
```

```python
#Feature(Petal Length) Vs Feature(Petal Width)
plt.scatter(data.iloc[0:50,3],data.iloc[0:50,2],c='r')
plt.scatter(data.iloc[50:100,3],data.iloc[50:100,2],c='b')
plt.scatter(data.iloc[100:150,3],data.iloc[100:150,2],c='y')
plt.xlabel("Petal Width")
plt.ylabel("Petal Length")
plt.title("Petal Length Vs Petal Width")
plt.legend(labels=['setosa','versicolor','virginica'])
plt.show()
```

```python
#Calculating mean of each class and each feature
slas=np.mean(list(data.iloc[0:30,0])) #sepal_length_avg_setosa
slavs=np.mean(list(data.iloc[50:80,0])) #sepal_length_avg_versicolor
slavg=np.mean(list(data.iloc[100:130,0]))#sepal_length_avg_virginica

swas=np.mean(list(data.iloc[0:30,1])) #sepal_width_avg_setosa
swavs=np.mean(list(data.iloc[50:80,1])) #sepal_width_avg_versicolor
swavg=np.mean(list(data.iloc[100:130,1]))#sepal_width_avg_virginica

plas=np.mean(list(data.iloc[0:30,2])) #petal_length_avg_setosa
plavs=np.mean(list(data.iloc[50:80,2])) #petal_length_avg_versicolor
plavg=np.mean(list(data.iloc[100:130,2]))#petal_length_avg_virginica

pwas=np.mean(list(data.iloc[0:30,3])) #petal_width_avg_setosa
pwavs=np.mean(list(data.iloc[50:80,3])) #petal_width_avg_versicolor
pwavg=np.mean(list(data.iloc[100:130,3]))#petal_width_avg_virginica
```
