My first kaggle analysis experience.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Preparing the data for calculating the correlation between all variables.

```python
df_students = pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
```

```python
df_correlation = pd.DataFrame(data={
    'genero': df_students['gender'].replace({'male':1, 'female':2}),
    'etnia': df_students['race/ethnicity'].replace({'group A': 1, 'group E': 2, 'group B': 3, 'group D': 4, 'group C': 5}),
    'escolaridade parental': df_students['parental level of education'].replace({
        "bachelor's degree": 5,
        'some college': 3,
        "master's degree": 6,
        "associate's degree": 4,
        'high school': 2,
        'some high school': 1}),
    'almoco': df_students['lunch'].replace({'free/reduced':1, 'standard': 2}),
    'curso preparatorio': df_students['test preparation course'].replace({'none': 1, 'completed': 2}),
    'matematica': df_students['math score'],
    'leitura': df_students['reading score'],
    'escrita': df_students['writing score'],
    'nota media': ((df_students['math score'] + df_students['reading score'] + df_students['writing score'])/3).astype(int)
})

df_students['p.l.e'] = df_students['parental level of education'].replace({
    "bachelor's degree": 5, "some college": 3, "master's degree": 6,
    "associate's degree": 4,'high school': 2, 'some high school': 1
})
```


* There are weak correlations between the categories 'Gender', 'Parental education', 'Lunch' and 'Preparatory course'.

```python
corr = df_correlation.corr()[['matematica','escrita','leitura','nota media']]
plt.figure(figsize=(5, 5))
plt.title("Correlation")
heatmap = sns.heatmap(corr[:5], annot=True)
```

* There is a strong correlation between writing and reading skills

```python
plt.figure(figsize=(5, 5))
plt.title("Correlação")
heatmap = sns.heatmap(corr[5:], annot=True)
```

* Creating function to facilitate the plotting of graphics.

```python
def plotGrid(x, y, data):
    f = plt.figure(figsize=(12,8))
    g = f.add_gridspec(2, 2)

    f.add_subplot(g[0,0])
    boxplot = sns.boxplot(x=x[0], y=y, data=data, palette="mako")

    f.add_subplot(g[0,1])
    boxplot = sns.boxplot(x=x[1], y=y, data=data, palette="mako")

    f.add_subplot(g[1,0])
    boxplot = sns.boxplot(x=x[2], y=y, data=data, palette="mako")

    f.add_subplot(g[1,1])
    boxplot = sns.boxplot(x=x[3], y=y, data=data, palette="mako")
```

According to the available database and evaluating the relationship of the variables with the written note:

* The female gender seems to have greater ease with escita than the male gender.
* Students with a more regular lunch have a slight advantage over students who do not have regular lunch.
* Students who have taken the preparatory course have an advantage over students who do not.
* Students whose parents are highly graduated tend to concentrate on higher grades.

```python
plotGrid(['gender','lunch','test preparation course','p.l.e'], 'writing score', df_students)
```

Students' reading performance is similar to their writing skills, with a few points:

* Students with highly graduated parents have slightly greater range of grades compared to their written grades.

```python
plotGrid(['gender','lunch','test preparation course','p.l.e'], 'reading score', df_students)
```

Students' reading performance is similar to their writing skills, with a few points:

* Students of different genders now have closer grades, with a slight improvement in males.
* The breadth of grades is less regardless of parents' education.

```python
plotGrid(['gender','lunch','test preparation course','p.l.e'], 'math score', df_students)
```
