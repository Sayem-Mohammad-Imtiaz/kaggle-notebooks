```python
## Accidents by City
Accidents_by_city=df.City.value_counts()
Accidents_by_city
```

```python
High_Accidents_Cities =Accidents_by_city[Accidents_by_city >=5000]
High_Accidents_Cities
```

```python
### Top 5 Cities that has accidents more than 5000
High_Accidents_Cities[:6].plot(kind='line')
```

```python
## Top 5 cities that has accidents less than 1000
Low_Accident_Cities = Accidents_by_city[Accidents_by_city <=1000]
Low_Accident_Cities.head(5)
```

```python
## Plotting top 5 cities that has accidents less than 1000
Low_Accident_Cities[:6].plot(kind='line')
### Hence we infer that safest cities in terms of accidents across the US where the number of accidents were less than 1000 are
###Edmond, The Dalles, Selma, Royal Oak, Lakeland
```

```python
## What time of the day does most accidents happened?
## What week of the dat records more accidents?
### Which months have the most accidents?
## What is the trend of the accidents year over year?
```

```python
df.Start_Time[0].hour
```

```python
### Here I copied the dataframe as an extra precaution to preserve the original dataframe and analysis done till this point
df1 = df.copy()
df1['hour'] = df['Start_Time'].dt.hour
```

```python
df1[['Start_Time','hour']][:15]
```

```python
### Plotting Accident data per hour
import seaborn as sns
sns.countplot(x="hour", data=df1)
### Here we can infer that accidents are on a rise between 5 AM to 6 PM
```

```python
df1['Year'] = df['Start_Time'].dt.year
```

```python
df1[['Start_Time','Year']]
```

```python
## Plotting Accidents per year
sns.countplot(x="Year", data=df1)
```

```python
##Getting the weekday from Start_Time
df1['Weekday'] = df1.Start_Time.dt.day_name()
```

```python
df1[['Start_Time','Weekday']]
```

```python
## Plotting Accidents per weekday
sns.countplot(x="Weekday", data=df1)
```

```python
df1['Month'] = df1.Start_Time.dt.month_name()
```

```python
df1[['Start_Time','Month']]
```

```python
##Plotting Accidents per month
sns.countplot(x="Month", data=df1)
```
