Since [we found out how age affects the number of uncommon foods a person eats](https://www.kaggle.com/bsoyka3/eating-more-foods-as-you-grow-older), I'm curious to know how other demographics affect this.

Start with some imports and the dataset.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("../input/have-you-ever-eaten-these-foods/responses.csv")
```

Now, we'll get the number of uncommon foods each respondent has had.

```python
foods = list(df)[4:]

def count_foods(row):
    total = 0
    
    for column in foods:
        if row[column]:
            total += 1
            
    return total

df["Foods eaten"] = df.apply(count_foods, axis=1)
```

Drop the columns we aren't using.

```python
df.drop(foods, inplace=True, axis=1)
df.drop("Age in years", inplace=True, axis=1)
```

Below are the results as bar graphs!

```python
gender = df.groupby("Gender").mean()
px.bar(x=gender.index, y=gender["Foods eaten"], title="Average foods eaten by gender").show()
```

```python
race = df.groupby("Race").mean()
px.bar(x=race.index, y=race["Foods eaten"], title="Average foods eaten by race").show()
```

```python
home_loc = df.groupby("Home location").mean()
px.bar(x=home_loc.index, y=home_loc["Foods eaten"], title="Average foods eaten by home location").show()
```
