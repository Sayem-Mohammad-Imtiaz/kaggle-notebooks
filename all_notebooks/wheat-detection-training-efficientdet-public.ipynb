```python
from omegaconf import OmegaConf, DictConfig

from wheat.train import train
from wheat.config import WheatConfig
```

```python
config = WheatConfig(
    base_dir="/kaggle/input/global-wheat-detection-dataset-resized-to-768x768/768/",
    # base_dir="/kaggle/input/wheat-dataset-resized-to-512x512/512/",
    epochs=16,
    image_size=768,
    arch="tf_efficientdet_d3",
    grad_accu=8,
    batch_size=4,
    precision=32,
    base_lr=4e-4,
    cutout=True,
    mixup_alpha=8.,
    seed=6321,
    no_op_ratio=0.4,
    no_op_warmup_steps=2000,
    mosaic_p=0.8
)
omega_conf = OmegaConf.structured(config)
with open("train_config.yaml", "w") as fout:
    OmegaConf.save(config=omega_conf, f=fout)
train(omega_conf, fold=0)
```
