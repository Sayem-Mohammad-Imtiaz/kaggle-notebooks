```python
data = pd.read_csv('/kaggle/input/world-happiness-report-2021/world-happiness-report.csv')
data_2021 = pd.read_csv('/kaggle/input/world-happiness-report-2021/world-happiness-report-2021.csv')
data_2021['year'] = 2021
data_2021 = data_2021[['Country name','year','Ladder score', 'Logged GDP per capita', 'Social support', 'Healthy life expectancy',
         'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']]
data_2021.columns = data.columns[:-2]

data = data.append(data_2021, ignore_index = True)
data['year'] = pd.to_datetime(data['year'], format = "%Y")
data['Log GDP per capita'] = np.exp(data['Log GDP per capita'])

scaler = MinMaxScaler()
data[data.columns[2:]] = scaler.fit_transform(data[data.columns[2:]])

data_br = data[data['Country name'] == 'Brazil']

timeline = pd.read_csv('../input/brazilliantimeline/brazil-timeline.csv')
timeline['date'] = pd.to_datetime(timeline['date'])
data_br['Log GDP per capita'] = data_br['Log GDP per capita']*6
```

```python
template = "plotly_dark"

fig1 = px.line(data_br, x = 'year', y = data_br.columns[2:], template = template)
fig1.update_traces(hovertemplate = "<br><b>%{y}</b>")

fig1.update_traces(visible = 'legendonly')

fig2 = px.scatter(timeline, x = 'date', y = np.ones(len(timeline)), hover_data = ['rulling_party','event'], symbol = 'category',
             color = 'rulling_party', template = template)

fig2.update_traces(hovertemplate = "<br>%{customdata[1]} <br>%{x}")
for trace in fig2.data:
    fig1.add_trace(trace)

fig1.update_traces(marker=dict(size=15, line=dict(width=1, color='black')), selector=dict(mode='markers'), textposition='top center')

fig1.update_layout(hovermode="x unified")


fig1.update_layout(
    height=500,
    width = 860)


fig1.show()
```
