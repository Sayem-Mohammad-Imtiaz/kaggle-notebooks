```python
confirmed_cases = pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/covid_19_data.csv')
```

```python
death_reported = pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/time_series_covid_19_deaths.csv')
```

```python
recovered_cases = pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/time_series_covid_19_recovered.csv')
```

```python
confirmed = confirmed_cases.loc[ cols[4]:cols[-1]]
```

```python
death = death_reported.loc[cols[4]: cols[-1]]
```

```python
recovered = recovered_cases.loc[cols[4]: cols[-1]]
```

```python
#finding the total confirmed cases, deatgh cases and the recovered cases and append them to an 4 empty lists.
#Also, calculate the total mortality rate which is the death_sum/confirmed_cases.

dates = confirmed.keys()
world_cases = []
total_cases = []
mortality_ratio = []
total_recovered = []

for i in dates:
    confirmed_sum = confirmed[i].sum()
    death_sum = death[i].sum()
    recovered_sum = recovered[i].sum()
    world_cases.append(confirmed_sum)
    total_cases.append(death_sum)
    mortality_ratio.append(death_sum/confirmed_sum)
    total_recovered.append(recovered_sum)
    
```
