# Support Vector Macchine (SVR)

## Importing the Libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # for Visualizing the data

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

## Importing the dataset

```python
filepath = '../input/heightvsweight-for-linear-polynomial-regression/HeightVsWeight.csv'
dataset = pd.read_csv(filepath)

features = ['Age']
label = ['Height']

x = dataset.loc[:, features].values
y = dataset.loc[:, label].values
```

## Splitting the dataset into training and test sub-datasets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
y_train = sc_y.fit_transform(y_train)
#y_test = sc_y.transform(y_test)
```

## Training SVR Model

```python
from sklearn.svm import SVR
reg = SVR(kernel = 'rbf')
reg.fit(x_train, y_train)
```

## Plotting the Graph for SVR

```python
plt.scatter(sc_x.inverse_transform(x_train), sc_y.inverse_transform(y_train), color = 'red')
plt.plot(sc_x.inverse_transform(x_test),sc_y.inverse_transform(reg.predict(x_test)), color = 'blue')
plt.title('SVR')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.show()
```

## Comparing Actual and Predicted results with SVR 

```python
np.set_printoptions(precision = 2) # for printing only upto 2 decimal places
y_pred = sc_y.inverse_transform(reg.predict(x_test))
print(np.concatenate((y_test.reshape(len(y_test), 1), y_pred.reshape(len(y_pred), 1)), 1))
```
