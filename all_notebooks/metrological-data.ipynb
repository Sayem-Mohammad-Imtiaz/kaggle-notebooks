```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/weather-dataset/weatherHistory.csv')
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Precip Type'] = df['Precip Type'].replace(np.nan,method='ffill')
```

```python
df['Precip Type'].unique()
```

```python
df['Precip Type'] = df['Precip Type'].astype('category').cat.codes
```

```python
df['Precip Type'].unique()
```

```python
df.columns
```

```python
df.drop(columns=['Loud Cover','Pressure (millibars)','Daily Summary'],axis=1,inplace=True)
```

```python
df.head()
```

```python
import datetime
df['Date'] = pd.to_datetime(df['Formatted Date'],utc=True)
df['Month'] = df['Date'].dt.month
```

```python
df.head()
```

```python
df.drop(['Formatted Date'],axis=1,inplace=True)
```

```python
df['Summary'].value_counts()
```

```python
from sklearn import preprocessing
label = preprocessing.LabelEncoder()
df['Summary'] = label.fit_transform(df['Summary'])
df['Summary'].unique()
```

```python
req_features = df[['Date','Apparent Temperature (C)','Humidity']]
```

```python
req_features.head()
```

```python
def year(fdate):
    x=str(fdate).split(':')[0][:-3].split('-')
    return int(x[0])
def month(fdate):
    x=str(fdate).split(':')[0][:-3].split('-')
    return int(x[1])
def day(fdate):
    x = str(fdate).split(':')[0][:-3].split('-')
    return int(x[2])
```

```python
req_features['Year'] = req_features['Date'].apply(year)
req_features['Month'] = req_features['Date'].apply(month)
req_features['Day'] = req_features['Date'].apply(day)
req_features.drop('Date',axis=1,inplace=True)
req_features.head()
```

```python
req_features['Year'].unique()
```

```python
plt.scatter('Apparent Temperature (C)','Humidity',data=req_features,marker='o');
```

```python
df_2016=req_features[req_features['Year']==2016]
df_2012=req_features[req_features['Year']==2012]
df_2010=req_features[req_features['Year']==2010]
df_2008=req_features[req_features['Year']==2008]
df_2015=req_features[req_features['Year']==2015]
df_2013=req_features[req_features['Year']==2013]
df_2011=req_features[req_features['Year']==2011]
df_2009=req_features[req_features['Year']==2009]
df_2007=req_features[req_features['Year']==2007]
df_2006=req_features[req_features['Year']==2006]
df_2014=req_features[req_features['Year']==2014]
```

```python
df_list = [df_2006,df_2007,df_2008,df_2009,df_2010,df_2011,df_2012,df_2013,df_2014,df_2015,df_2016]
max_temp = []
min_temp = []
avg_temp = []
max_humidity = []
min_humidity = []
avg_humidity = []
for i in df_list:
    max_temp.append(i['Apparent Temperature (C)'].max())
    min_temp.append(i['Apparent Temperature (C)'].min())
    avg_temp.append(i['Apparent Temperature (C)'].mean())
    max_humidity.append(i['Humidity'].max())
    min_humidity.append(i['Humidity'].min())
    avg_humidity.append(i['Humidity'].mean())
    
```

```python
year_list=[2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016]
plt.figure(figsize=(12,6))
sns.lineplot(data=pd.DataFrame({'Years':year_list,'Max Temparature':max_temp,'Min Temparature':min_temp,'Average Temparature':avg_temp}).set_index('Years'))
plt.figure(figsize=(12,6))
sns.lineplot(data=pd.DataFrame({'Years':year_list,'Max Humidity':max_humidity,'Min Humidity':min_humidity,'Average Humidity':avg_humidity}).set_index('Years'))
```

```python
x = req_features.iloc[:,:2].values
y = req_features.iloc[:,2:4].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=20,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
print('Coefficient:\n',lr.coef_)
print('Mean Square error:%.2f'%mean_squared_error(y_test,y_pred))
```
