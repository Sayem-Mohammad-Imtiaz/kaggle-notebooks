```python
import random
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fpl = pd.read_csv(r'../input/fpl-player-data-2021/FPL 20 21.csv')
fpl.head()
```

```python
Player = fpl.Second_Name
GI = fpl.Goals_Scored + fpl.Assists
df = pd.DataFrame({"Player":Player, "Goal_Involvements":GI}).sort_values(by='Goal_Involvements', ascending = False)[:20]

df=df.reset_index(drop=True)
n = df['Player'].unique().__len__()+1
all_colors = list(plt.cm.colors.cnames.keys())
random.seed(100)
c = random.choices(all_colors, k=n)
plt.figure(figsize=(16,5), dpi= 80)
plt.bar(df['Player'], df['Goal_Involvements'], color=c, width=.5)
for i, val in enumerate(df['Goal_Involvements'].values):
    plt.text(i, val, int(val), horizontalalignment='center', verticalalignment='bottom', fontdict={'fontweight':500, 'size':14})
plt.gca().set_xticklabels(df['Player'], rotation=40, horizontalalignment= 'center', size=11)
plt.title("Total Goal Involvements", fontsize=22)
plt.ylabel('Goals + Assists',fontsize=15)
plt.ylim(0, 45)
plt.show()
```

```python
ppm_90 = []
player = []
for i in range(len(fpl)):
    if fpl.Minutes[i]>1800:
        ppm_90.append((fpl.Total_Points[i]/fpl.Minutes[i])*90)
        player.append(fpl.Second_Name[i])
df = pd.DataFrame({"Player":player, "Points_per_90":ppm_90}).sort_values(by='Points_per_90', ascending =False)[:20]

fig, ax = plt.subplots(figsize=(16,5), dpi= 100)
ax.vlines(x=df.Player, ymin=0, ymax=df.Points_per_90, color='blue', alpha=0.9, linewidth=3)
ax.scatter(x=df.Player, y=df.Points_per_90, s=20, color='red', alpha=1)
ax.set_title('Points per 90 Min', fontdict={'size':22})
ax.set_ylabel('Points / 90 Min',size=12)
ax.set_xlabel('Played over 1800 min',size=10)
ax.set_xticks(df.Player)
ax.set_xticklabels(df.Player, rotation=45, fontdict={'horizontalalignment': 'center', 'size':12})
ax.set_ylim(0, 10)

for row in df.itertuples():
    ax.text(row.Player, row.Points_per_90, s=round(row.Points_per_90, 2), horizontalalignment= 'center', verticalalignment='bottom', fontsize=14)

plt.show()
        
```

```python
team = fpl['Club'].unique()
first=[]
score=[]
second=[]
score_2=[]
for i in team:
    max=0
    player=""
    max_2=0
    player_2=""
    pos=0
    for k in range(len(fpl)):
        if (fpl.Club[k]==i and fpl.Total_Points[k]>max):
            max=fpl.Total_Points[k]
            pos=k
            player=fpl.Second_Name[k]
    first.append(player)       
    score.append(max)
    
    for j in range(len(fpl)):
        if (fpl.Club[j]==i and fpl.Total_Points[j]>max_2 and fpl.Total_Points[j]!=fpl.Total_Points[pos]):
            max_2=fpl.Total_Points[j]
            player_2=fpl.Second_Name[j]
                    
    second.append(player_2)       
    score_2.append(max_2)
    
df_first=pd.DataFrame({"First":first,"score":score})
df_second=pd.DataFrame({"Second":second,"score_2":score_2})

barWidth=0.25

bars1=[]
bars2=[]

for i in df_first.score: 
    bars1.append(i)
for i in df_second.score_2: 
    bars2.append(i)

r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]

plt.figure(figsize=(28,13))

plt.bar(r1, bars1, color='red', width=barWidth, edgecolor='black', label='1st')
plt.bar(r2, bars2, color='blue', width=barWidth, edgecolor='black', label='2nd')

for i in range(len(df_first['First'])):
    plt.text(i,df_first.score[i],str(df_first.First[i]), horizontalalignment='center',rotation=60, verticalalignment='bottom', fontdict={'fontweight':500, 'size':13,'fontsize':17,'color':'red'})
 


for i in range(len(df_second['Second'])):
    plt.text(i,df_second.score_2[i],str(df_second.Second[i]),horizontalalignment='left',rotation=60, verticalalignment='bottom', fontdict={'fontweight':500, 'size':13,'fontsize':17,'color':'blue'})


plt.xticks([r + 0.125 for r in range(len(bars1))], ['MUN', 'TOT', 'LIV', 'LEE', 'LEI',
       'AVL', 'EVE', 'MCI', 'SOU', 'WHU',
       'WBA', 'CHE', 'BUR', 'CRY', 'NEW',
       'WOL', 'BHA', 'ARS', 'FUL', 'SHU'],fontsize=19)
 
plt.ylabel("Points",fontsize=20)
plt.legend(fontsize=25)
plt.title("Top 2 (Points)",fontsize=35)
plt.show()
```
