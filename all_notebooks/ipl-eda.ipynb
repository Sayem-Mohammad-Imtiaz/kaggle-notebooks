```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Matches dataset

```python
matches_df = pd.read_csv('../input/ipl-data/matches.csv', index_col = 0)
matches_df.head()
```

### Top 10 Cities Matches played in.

Seems like Mumbai was most prefered city to play match, followed by Bangalore and Kolkata. 

### Season wise numbr of matches played

2013 has most number of matches played in that too above 70.

### Team with most wins

```python
data = matches_df.winner.value_counts()
sns.barplot(y = data.index, x = data, orient='h')
```

As you can see Mumbai Indians tops the chart here.

### Stadium wise matches

### Toss decision

```python
temp_series = matches_df.toss_decision.value_counts()
labels = (np.array(temp_series.index))
sizes = (np.array((temp_series / temp_series.sum())*100))
plt.pie(sizes, labels=labels,autopct='%1.1f%%');
```

Most of the time after winning the toss teams seems to prefer choose field first.

```python
sns.countplot(x='season', hue='toss_decision', data=matches_df)
```

We can say since 2014 teams started to prefer field instaed of batting after winning the toss.

### Choosing Fielding first win percentage

```python
num_of_wins = (matches_df.win_by_wickets>0).sum()
num_of_loss = (matches_df.win_by_wickets==0).sum()
labels = ["Wins", "Loss"]
total = float(num_of_wins + num_of_loss)
sizes = [(num_of_wins/total)*100, (num_of_loss/total)*100]
plt.pie(sizes, labels=labels, autopct='%1.1f%%');
```

From above plot we can say when team choose to field first won 53 times out of 100.

### Deliveries Dataset

```python
deliver_df = pd.read_csv('../input/ipl-data/deliveries.csv', index_col=0)
deliver_df.head()
```

```python
deliver_df.shape
```

### Batsman Analysis

Let's see which batsman has scored most runs in all the season combined.

```python
temp_df = deliver_df.groupby('batsman')['batsman_runs'].agg('sum').sort_values(ascending = False)
temp_df
```

```python
runs = sns.barplot(x = temp_df[:10], y = temp_df[:10].index, orient = 'h')
```

```python
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 1.02*height,
                '%d' % int(height),
                ha='center', va='bottom')
```

```python
labels = np.array(temp_df[:10].index)
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(10,7))
rects = ax.bar(ind, np.array(temp_df[:10]), width=width, color='blue')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top run scorers in IPL")
ax.set_xlabel('Batsmane Name')
autolabel(rects)
plt.show()
```

Suresh Raina(Chinna Thala) is leading the chart followed closely by Current Indian team captain King Kohli. David Warner(Tik-tok star) is the top scorer among foreign players.

### Numbers of 4s
let us see the players with more number of 4s in IPL.

```python
temp_df = deliver_df.groupby('batsman')['batsman_runs'].agg(lambda x :(x==4).sum()).sort_values(ascending = False)
temp_df
```

```python
labels = np.array(temp_df[:10].index)
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(10,7))
rects = ax.bar(ind, np.array(temp_df[:10]), width=width, color='blue')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top run scorers in IPL")
ax.set_xlabel('Batsmane Name')
autolabel(rects)
plt.show()
```

Gambhir is way ahead of others - almost 80 boundaries more than Raina.! Nice to Sehwag in the top 10 list :)

### Numbers of 6s

Now let us check the number of 6's

```python
temp_df = deliver_df.groupby('batsman')['batsman_runs'].agg(lambda x :(x==6).sum()).sort_values(ascending = False)


labels = np.array(temp_df[:10].index)
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(10,7))
rects = ax.bar(ind, np.array(temp_df[:10]), width=width, color='blue')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top run scorers in IPL")
ax.set_xlabel('Batsmane Name')
autolabel(rects)
plt.show()
```

As expected the super sixer king Chris the Storm Gayle tops the chart with the whooper 266 sixes. Followed by Raina and Rohit Sharma difference between them is only of 1 six. Raina is always 2nd in number of fours and sixes.

### Number of Dot balls
Now let us see the batsman who has played the most number of dot balls.

```python
temp_df = deliver_df.groupby('batsman')['batsman_runs'].agg(lambda x :(x==0).sum()).sort_values(ascending = False)

labels = np.array(temp_df[:10].index)
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(10,7))
rects = ax.bar(ind, np.array(temp_df[:10]), width=width, color='blue')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top run scorers in IPL")
ax.set_xlabel('Batsmane Name')
autolabel(rects)
plt.show()
```

It is interesting to see that the same names repeat again here as well. I think since these guys have played more number of balls, they have more dot balls as well.

## Bowler Analysis
Now let us see the bowlers who has bowled most number of balls in IPL.

```python
temp_df = deliver_df.groupby('bowler')['ball'].agg('count').sort_values(ascending=False)

labels = np.array(temp_df[:10].index)
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(10,7))
rects = ax.bar(ind, np.array(temp_df[:10]), width=width, color='blue')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top Bowlers - Number of balls bowled in IPL")
ax.set_xlabel("Bowler Name")
autolabel(rects)
plt.show()
```

Harbhajan Singh is the the bowler with most number of balls bowled in IPL matches. If you see he has been playing since 2008 season till now.

### Number of dot balls
let us see the bowler with more number of dot balls.

```python
temp_df = deliver_df.groupby('bowler')['total_runs'].agg(lambda x: (x==0).sum()).sort_values(ascending=False)


labels = np.array(temp_df[:10].index)
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(10,7))
rects = ax.bar(ind, np.array(temp_df[:10]), width=width, color='blue')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top Bowlers - Number of dot balls bowled in IPL")
ax.set_xlabel("Bowler Names")
autolabel(rects)
plt.show()
```

Pravin Kumar is the one with more number of dot balls followed by Bhajji and Slinga Malinga.

### Extra runs by which bowler

Let us see the bowlers who has bowled more number of extras in IPL.

```python
temp_df = deliver_df.groupby('bowler')['total_runs'].agg(lambda x: (x>0).sum()).sort_values(ascending=False)

labels = np.array(temp_df[:10].index)
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots(figsize=(10,7))
rects = ax.bar(ind, np.array(temp_df[:10]), width=width, color='blue')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top Bowlers - Number of dot balls bowled in IPL")
ax.set_xlabel("Bowler Names")
autolabel(rects)
plt.show()
```

Again we see the name of Bhajji on the top of the chart. Also if you see he is followed by Amit Mishra and Piyush Chawla both spinner.
