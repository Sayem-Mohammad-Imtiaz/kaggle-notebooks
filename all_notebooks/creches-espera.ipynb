```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../input/tempoatendimentocrechessp-2/tempo_de_atendimento_creches_prefeiturasp_2020.csv")
df
```

```python
df = df.drop(labels=0, axis=0)
df
```

```python
df.columns = ["Distrito", "Tempo(dias)"]
df
```

```python
x = df["Distrito"]
y = df["Tempo(dias)"]
x_arange = np.arange(len(x))


plt.title("Espera por vagas em creches")
plt.xlabel("Distritos")
plt.ylabel("Espera(dias)")

# plt.xticks(x_arange, x)
plt.bar([numero+1 for numero in x_arange], y, width=0.7)
```
