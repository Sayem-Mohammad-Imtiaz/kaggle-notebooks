```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
covid = pd.read_csv('../input/novel-covid-data-till-august-2021/covid-data.csv')
```

```python
covid.head()
```

**Every Country in Every Continent**

```python
continent_wise = covid.groupby(['continent','location'])['total_cases'].max().reset_index()
continent_wise
```

```python
africa = continent_wise[continent_wise['continent'] == 'Africa']
asia = continent_wise[continent_wise['continent'] == 'Asia']
europe = continent_wise[continent_wise['continent'] == 'Europe']
s_america = continent_wise[continent_wise['continent'] == 'South America']
n_america = continent_wise[continent_wise['continent'] == 'North America']
oceania = continent_wise[continent_wise['continent'] == 'Oceania']
```

```python
fig = px.bar(africa,y='location',x='total_cases',orientation='h', title='Total COVID-19 Cases in Each Country of Africa',height=1500)
fig.show()
```

```python
fig = px.bar(asia,y='location',x='total_cases',orientation='h', title='Total COVID-19 Cases in Each Country of Asia',height=1500)
fig.show()
```

```python
fig = px.bar(europe,y='location',x='total_cases',orientation='h', title='Total COVID-19 Cases in Each Country of Europe',height=1500)
fig.show()
```

```python
fig = px.bar(s_america,y='location',x='total_cases',orientation='h', title='Total COVID-19 Cases in Each Country of South America',height=700)
fig.show()
```

```python
fig = px.bar(n_america,y='location',x='total_cases',orientation='h', title='Total COVID-19 Cases in Each Country of North America',height=1000)
fig.show()
```

```python
fig = px.bar(oceania,y='location',x='total_cases',orientation='h', title='Total COVID-19 Cases in Each Country of Oceania',height=700)
fig.show()
```

**Cumulative Cases in Each Continent**

```python
most_cases = covid.groupby('location')['total_cases'].max().reset_index()
most_cases = most_cases.sort_values('total_cases', ascending=False)
most_cases.head(10)
```

```python
asia_cases = most_cases[most_cases['location'] == 'Asia']
europe_cases = most_cases[most_cases['location'] == 'Europe']
africa_cases = most_cases[most_cases['location'] == 'Africa']
s_america_cases = most_cases[most_cases['location'] == 'South America']
n_america_cases = most_cases[most_cases['location'] == 'North America']
world_cases = most_cases[most_cases['location'] == 'World']
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=asia_cases['location'], y=asia_cases['total_cases'], showlegend=False, name='Asia'))
fig.add_trace(go.Bar(x=europe_cases['location'], y=europe_cases['total_cases'], showlegend=False, name='Europe'))
fig.add_trace(go.Bar(x=n_america_cases['location'], y=n_america_cases['total_cases'], showlegend=False, name='North America'))
fig.add_trace(go.Bar(x=s_america_cases['location'], y=s_america_cases['total_cases'], showlegend=False, name='South America'))
fig.add_trace(go.Bar(x=world_cases['location'], y=world_cases['total_cases'], showlegend=False, name='Worldwide'))
fig.add_trace(go.Bar(x=africa_cases['location'], y=africa_cases['total_cases'], showlegend=False, name='Africa'))

fig.update_layout(title='Total Cases of COVID-19 in Each Continent', height=500)
fig.show()
```

**Most Affected Countries in the World**

```python
top_cases = covid.groupby('location')['total_cases'].max().reset_index()
top_cases = top_cases.sort_values('total_cases', ascending=False)
top_cases = top_cases.drop(top_cases.index[0:4])
top_cases = top_cases.drop(top_cases.index[1])
top_cases = top_cases.drop(top_cases.index[4])
top_cases = top_cases.drop(top_cases.index[1])
top_cases_10 = top_cases.head(10)
```

```python
fig = px.bar(top_cases_10, x='location',y='total_cases', title='COVID-19 Cases: Worst Affected Countries')
fig.show()
```

**Worst Affected Countries in terms of Total Deaths**

```python
top_10 = covid.groupby('location')['total_deaths'].max().reset_index()
deaths = top_10.sort_values('total_deaths',ascending=False)
deaths = deaths.drop(deaths.index[0:5])
most_deaths = deaths.iloc[0:12]
most_deaths = most_deaths.drop(most_deaths.index[0])
most_deaths = most_deaths.drop(most_deaths.index[6])
most_deaths
```

```python
fig=px.bar(most_deaths,x='location',y='total_deaths',title='COVID-19 Deaths')
fig.show()
```

**COVID-19 Deaths: Continent-Wise**

```python
continent_wise_deaths = covid.groupby('location')['total_deaths'].max().reset_index()
continent_wise_deaths = continent_wise_deaths.sort_values('total_deaths', ascending=False)
continent_wise_deaths.head(10)
```

```python
africa_deaths = continent_wise_deaths[continent_wise_deaths['location'] == 'Africa']
asia_deaths = continent_wise_deaths[continent_wise_deaths['location'] == 'Asia']
europe_deaths = continent_wise_deaths[continent_wise_deaths['location'] == 'Europe']
s_america_deaths = continent_wise_deaths[continent_wise_deaths['location'] == 'South America']
n_america_deaths = continent_wise_deaths[continent_wise_deaths['location'] == 'North America']
world_deaths = continent_wise_deaths[continent_wise_deaths['location'] == 'World']
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=africa_deaths['location'],y=africa_deaths['total_deaths'], name='Africa',showlegend=False))
fig.add_trace(go.Bar(x=asia_deaths['location'],y=asia_deaths['total_deaths'], name='Asia',showlegend=False))
fig.add_trace(go.Bar(x=europe_deaths['location'],y=europe_deaths['total_deaths'], name='Europe',showlegend=False))
fig.add_trace(go.Bar(x=s_america_deaths['location'],y=s_america_deaths['total_deaths'], name='South America',showlegend=False))
fig.add_trace(go.Bar(x=n_america_deaths['location'],y=n_america_deaths['total_deaths'], name='North America',showlegend=False))
fig.add_trace(go.Bar(x=world_deaths['location'],y=world_deaths['total_deaths'], name='Worldwide',showlegend=False))

fig.update_layout(title='Total COVID-19 Deaths on Every Continent')
fig.show()
```
