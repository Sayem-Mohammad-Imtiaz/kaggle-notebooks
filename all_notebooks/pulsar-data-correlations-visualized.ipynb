## Pulsar data correlations: visualized
This notebook is intended for correlation visualizations that exist between the pulsar HTRU2 dataset. 

```python
# library imports
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as poly

pulsar_file = '../input/pulsar-dataset-htru2/HTRU_2.csv'
```

```python
# reading the csv file and making a dataframe
#
# ip = integrated profile
# dm = dm/snr
#
pulsar_file_df = pd.DataFrame(pd.read_csv(pulsar_file, header=None))
pulsar_file_df.columns = ['mean_ip', 'std_ip', 'kurtosis_ip', 'skewness_ip',
                          'mean_dm', 'std_dm', 'kurtosis_dm', 'skewness_dm',
                          'class']

# showing the initial five rows of the datafram
pulsar_file_df.head()
```

```python
# correlations, but the last columns `class` is not included
pulsar_file_df.drop(['class'], axis=1).corr()
```

```python
# visualizing the correlations through seaborn heatmap
plt.figure(figsize=(8,6))
_ = sns.heatmap(pulsar_file_df.drop(['class'], axis=1).corr(), annot=True)
```

```python
# visualizing the correlations throught pandas scatter_matrix
# _ = pd.plotting.scatter_matrix(pulsar_file_df.drop(['class'], axis=1), figsize=(15, 15), diagonal='kde')

# visualizing the correlations through seaborn pairplot
_ = sns.pairplot(pulsar_file_df, kind='scatter', hue='class', plot_kws=dict(ec='w'))
```
