## Creat new feature
### We combine the 'sibling\spouse' and 'parent\children' as 'family size'

```python
#Creating new family_size column
data_train['Family_Size']=data_train['SibSp']+data_train['Parch']
```

```python
data_train.sample(5)
```

```python
sns.barplot(x="Family_Size", y="Survived", hue="Sex", data=data_train);
```

# Feature extraction on the training set
## We then create the feature matrix using 'Pclass', 'Sex', and 'Family size'
### And the target  array is the 'Survived '

```python
train_x = data_train[['Pclass','Sex','Family_Size']]
train_x.sample(5)
```

```python
train_y = data_train['Survived']
train_y.sample(4)
```

## Encoding
###  We use sklearn encoder to convert the string to numbers (dummy variables)

```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
train_x["Sex"] = lb_make.fit_transform(train_x["Sex"])
```

## We first split this training set into new training and test data to check accuracy by fitting different models (classifiers)
## In this work, we used :
### Guassian Naive Bayes
### Random Forest
### Support Vector Classifier (SVC) 
### Gradient Boosting Classifier

```python
from sklearn.cross_validation import train_test_split
```

```python
##split the data into two 
xtrain, xtest, ytrain, ytest = train_test_split(train_x, train_y,
random_state=1)
```

## We import the models from sklearn

```python
from sklearn.naive_bayes import GaussianNB 
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
model = GaussianNB()
model1= SVC()
model2 = RandomForestClassifier()
model3 = GradientBoostingClassifier()
```

```python
from sklearn.metrics import accuracy_score  ###checks the accuracy of the model
```

```python
models = [model, model1, model2, model3]
for i in models:
    i.fit(xtrain, ytrain)
    ypred = i.predict(xtest) 
    print(i ,accuracy_score(ytest, ypred))
```

From the above test, the $\textbf{Gaussian Naive Bayes}$ is more accurate.<p>

We then apply it on the whole training set to make predictions 

## Fit Gaussian Naive Bayes on the training set

```python
model.fit(train_x, train_y)
```

# Prediction using the model

## We do the same feature engineering and extraction,   and encoding on the test data

```python
data_test['Family_Size']=data_test['SibSp']+data_test['Parch']
test_x = data_test[['Pclass','Sex','Family_Size']]
test_x["Sex"] = lb_make.fit_transform(test_x["Sex"])
test_x.sample(5)
```

## predicting the survived using the Gaussian Naive Bayes

```python
y_model = model.predict(test_x)
```

## To create a submission file (csv) using Pandas 

```python
Pa_id = data_test['PassengerId']
```

```python
results = pd.DataFrame({ 'PassengerId' : Pa_id, 'Survived': y_model})
results.sample(5)
```

```python
##Export as csv file
results.to_csv('titanic_pred_results.csv', index = False)
```
