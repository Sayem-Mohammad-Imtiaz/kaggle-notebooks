```python
def cat_a_num (columna):
    if columna == "single":
        return 0
    elif columna == "married":
        return 1
    else:
        return 2
```

```python
def education_tr (education):
    if education == "illiterate":
        return 0
    if education == "basic.4y":
        return 1
    if education == "basic.6y":
        return 2
    if education == "basic.9y":
        return 3
    if education == "high.school":
        return 4
    if education == "professional.course":
        return 5
    else:
        return 6 
```

```python
df["education_bin"] = df["education"].apply(education_tr)
```

```python
def convert_day (day_of_week):
    if day_of_week == "mon":
        return 0
    elif day_of_week == "tue":
        return 1
    elif day_of_week == "wed":
        return 2
    elif day_of_week == "thu":
        return 3
    else:
        return 4
```

```python
df["day_of_week_bin"] = df["day_of_week"].apply(convert_day)
```

```python
def job_bn (job):
    if job == "admin.":
        return 0
    if job == "blue-collar":
        return 1
    if job == "entrepreneur":
        return 2
    if job == "housemaid":
        return 3
    if job == "management":
        return 4
    if job == "retired":
        return 5
    if job == "self-employed":
        return 6
    if job == "services":
        return 7
    if job == "technician":
        return 8
    if job == "unemployed":
        return 9
    if job == "unknown":
        return 10
    else:
        return 11
```

```python
df["job"] = df["job"].apply(job_bn)
```

```python
def convert_age (age):
    if age <= 10:
        return 0
    elif 10 < age <= 20:
        return 1
    elif 20 < age <= 30:
        return 2
    elif 30 < age <= 40:
        return 3
    elif 40 < age <= 50:
        return 4
    elif 50 < age <= 60:
        return 5
    elif 60 < age <= 70:
        return 6
    elif 70 < age <= 80:
        return 7
    elif 80 < age <= 90:
        return 8
    else:
        return 9
```

```python
df["age_bin"] = df["age"].apply(convert_age)
```

```python
df_x = df.loc[:,["emp.var.rate","age", "euribor3m", "nr.employed", "duration", "job","cons.conf.idx", "cons.price.idx"]]
df_y = df['y_yes']
trainX, testX, trainY, testY = train_test_split(df_x,df_y,test_size = 0.8,    
                                            random_state=27)

from sklearn.linear_model import LogisticRegression

LR = LogisticRegression(solver='liblinear', random_state=27) 
LR.fit(trainX,trainY) 
y_prediccion = LR.predict(trainX)
```

```python
LR.score(trainX,trainY)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

classification_report(trainY, y_prediccion)
```

```python
conf_matrix = confusion_matrix(trainY,y_prediccion)
conf_matrix

```
