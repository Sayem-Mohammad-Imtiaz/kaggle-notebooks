```python
df.Income_Category = df.Income_Category.map({
                                            "Less than $40K":1,
                                            "$40K - $60K":2,
                                            "$80K - $120K":3,
                                            "$60K - $80K":4,
                                            "$120K +":5        
                                                })
```

```python
df.Card_Category = df.Card_Category.map({
    'Blue':1,
    'Silver':2,
    'Gold':3,
    'Platinum':4
})
```

```python
df_temp = df[df.Education_Level=='Unknown']
df_temp.shape
```

```python
df_temp.Attrition_Flag.value_counts()
```

```python
df = df[df.Education_Level!='Unknown']
```

```python
df.Education_Level = df.Education_Level.map(
{
'Uneducated':1,
'High School':2,
'College':3,
'Graduate':4, 
'Post-Graduate':5,
'Doctorate':6   
})
```

```python
df.head()
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df.columns
```

```python
x = df.drop(['Attrition_Flag'] , axis = 1)
y = df.Attrition_Flag
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.3 , random_state = 100)
```

```python
model = RandomForestClassifier(n_estimators = 100)
model.fit(x_train , y_train)
y_pred = model.predict(x_test)
print("Accuracy:",accuracy_score(y_test , y_pred))
arr = confusion_matrix(y_test , y_pred)
recall = arr[1][1]/(arr[0][1]+arr[1][1])
print("Recall:" , recall)
```

```python
for i in range(100 , 1000 , 200):
    model = RandomForestClassifier(n_estimators = i)
    model.fit(x_train , y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test , y_pred)
    arr = confusion_matrix(y_test , y_pred)
    recall = arr[1][1]/(arr[0][1]+arr[1][1])
    print(f"Accuracy: {accuracy} and recall: {recall} for Decision Tree={i}")
```
