```python
def kategorik_degerleri_duzenle(deger):
    if deger == "Var":
        return 1
    else:
        return 0
```

```python
df["Bluetooth"] = df["Bluetooth"].apply(kategorik_degerleri_duzenle)
df["CiftHat"] = df["CiftHat"].apply(kategorik_degerleri_duzenle)
df["4G"] = df["4G"].apply(kategorik_degerleri_duzenle)
df["3G"] = df["3G"].apply(kategorik_degerleri_duzenle)
df["Dokunmatik"] = df["Dokunmatik"].apply(kategorik_degerleri_duzenle)
df["WiFi"] = df["WiFi"].apply(kategorik_degerleri_duzenle)
```
