```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

# Reading the csv file

```python
df=pd.read_csv('/kaggle/input/indianeedsoxygen-tweets/IndiaWantsOxygen.csv')
```

# Trying to understand the data and reshaping it to get the usefull content

```python
df.head(10)
```

```python
df.isnull().sum()
```

**Dropping the username with null values**

```python
df= df[pd.notnull(df["user_name"])]
```

```python
df.head(3)
```

# Plotting the tweets in a black background in random fashion by using "TEXT"

```python
from wordcloud import WordCloud, STOPWORDS
plt.figure(figsize= (20,20))
words= "".join(df["text"])
final = WordCloud(width = 2000, height = 800, background_color ="black",min_font_size = 10).generate(words)
plt.imshow(final)
plt.show()
```

#  Plotting the tweets in a black background in random fashion by using "HASHTAGS"

```python
from wordcloud import WordCloud, STOPWORDS
plt.figure(figsize= (20,20))
words= "".join(df["hashtags"])
final = WordCloud(width = 2000, height = 800, background_color ="black",min_font_size = 10).generate(words)
plt.imshow(final)
plt.show()
```

**Checking the unique countries**

```python
df['user_location'].unique()
```

# Counting the number of tweets from top 15 countries

```python
top_tweets=df.user_location.value_counts().head(15)
top_tweets
```

# Plotting the bargraph of top 15 location having maximum tweets

```python
a=sns.barplot(top_tweets.values,top_tweets.index)
a.set_ylabel("Name of the location")
a.set_xlabel("Number of tweets")
plt.show()
```

# Converting the date into date format

```python
df["date"]= pd.to_datetime(df.date)
df.date= df.date.apply(lambda x: str(x).split(" ")[0])
df.date
```

# Plotting the maximum tweet in which date

```python
x= df.groupby("date").date.count().head(15)
sns.set_style("whitegrid")
b= sns.lineplot(x.index,x.values)
plt.figure(figsize= (15,7))
b.set_xlabel("date")
b.set_ylabel("No of tweets")
plt.show()
```
