```python
df = pd.read_csv("/kaggle/input/handwriting-recognition/written_name_train_v2.csv")
```

```python
X = []
Y = []
for i, (x, y) in df.iterrows():
    X.append(x)
    res = []
    if type(y) == float:
        continue
    for l in y:
        l_encoded = [1 if i == ord(l) - ord('A') else 0 for i in range(27)]
        res.append(l_encoded)
    res.append([0 for i in range(26)] + [1])
    Y.append(res)
```
