# 1. I selected India.

In India, a rapid increase in the number of patients has been reported. We checked the data to see how much it actually increased.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.set_printoptions(threshold=np.inf)

selected_country='India'
df = pd.read_csv('../input/novel-corona-virus-2019-dataset/covid_19_data.csv',header=0)
#print(np.unique(df['Country/Region'].values))

df = df[df['Country/Region']==selected_country]
df = df.groupby('ObservationDate').sum()
print(df)
```

# 2 As shown in the 3 figures below, I constructed a time-series and colored table regarding daily confirmed case, daily deaths and daily recoveries.

```python
df['daily_confirmed'] = df['Confirmed'].diff()
df['daily_deaths'] = df['Deaths'].diff()
df['daily_recovery'] = df['Recovered'].diff()
df['daily_confirmed'].plot()
df['daily_recovery'].plot()
plt.show()
```

```python
print(df)
```

# Figure of daily confirmed case, daily deaths and daily recoveries

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_confirmed_object = go.Scatter(x=df.index,y=df['daily_confirmed'].values,name='Daily confirmed')
daily_deaths_object = go.Scatter(x=df.index,y=df['daily_deaths'].values,name='Daily deaths')
daily_recovery_object = go.Scatter(x=df.index,y=df['daily_recovery'].values,name='Daily_recovery')


layout_object = go.Layout(title='Fig.1 India daily cases 19M52596',xaxis=dict(title='Date'),yaxis=dict(title='Number of people'))
fig = go.Figure(data=[daily_confirmed_object,daily_deaths_object,daily_recovery_object],layout=layout_object)
iplot(fig)
fig.write_html('India_daily_cases_19M52596.html')
```

# 4 Discussion about Corona Policy in India

In India, the first person to be infected with COVID19 was confirmed on January 30, 2019.
Since then, the Indian government has implemented countermeasures such as entry and travel restrictions in February and March. As a result, as shown in Figure.1, the number of infected people remains small. However, by the end of March, the number of infected people began to gradually increase, and lockdown countermeasures were taken in some areas.
In fact, when checking Table 1, it can be seen that the number of infected people began to rapidly increase around 3/20, and that they took precautionary measures against the rapid increase in the number of infected persons.

Although the number of infected people continued to increase, the explosion increase of the patients could be avoided by extending the lockdown measures, and the worst situation such as medical collapse could be prevented. On the other hand, as concerns about economic impact such as an increase in the number of unemployed people have begun to appear, the government of India has a policy to gradually permit economic activities from June, except for the "red zone" where the number of coronavirus cases is high. 

However, there were problems with this policy. As a result of permitting economic activities, as shown in Fig. 1, the number of patients continues to increase, and more than 20,000 infected people was confirmed, which greatly exceeds the peak value so far in the middle of June. Also, As a result, as you can see from the figure, the number of patients increased rapidly, and in about two months, the cumulative number of patients increased to the fourth largest in the world (as shown #3)
Such a rapid increase in the number of patients may lead to medical collapse and is extremely dangerous. 
Since it was expected that the number of unemployed people will increase due to the lockdown, it is considered that the budget for the unemployed people should have been prepared.
