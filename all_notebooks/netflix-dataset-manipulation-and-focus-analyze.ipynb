## Is Netflix has increasingly focusing on TV rather than movies in recent years ?


*Slicing the month names from the date column*

```python
df["date_added"] = df['date_added'].str.slice(start=-4)
df["date_added"]
```

*Dropping The NaN Values*

```python
df.dropna(inplace = True)
df
```

```python
print(df['date_added'].dtypes)
```

```python
df["date_added"].values.min()
```

```python
df["date_added"].values.max()
```

*Converting objects to int64*

```python
df["date_added"] = df["date_added"].astype("int64")
df["date_added"]
```

## **Partitioning to focus on recent years**

```python
year_cuts = pd.cut(df["date_added"],[2008,2015,2020])
year_cuts
```

```python
df["type"].isnull()
```

```python
from sklearn import preprocessing
le_type = preprocessing.LabelEncoder()
le_type.fit(['Movie','TV Show'])
df.iloc[:,1] = le_type.transform(df.iloc[:,1]) 
df.iloc[0:10]
```

```python
df.pivot_table("type",["date_added",year_cuts],"release_year",aggfunc="sum")
```

## Heatmapping with sum of values to see the answer that: Is netflix's focus on TV shows in recent years ?

```python
import seaborn as sns
from matplotlib import pyplot

fig, ax = pyplot.subplots(figsize=(18,10))
df_heatmap = df.pivot_table("type",["date_added",year_cuts],"release_year",aggfunc="sum")
sns.heatmap(df_heatmap,annot = True,ax = ax);
```

# Insights and result

*Tv shows coded as 1's , Movies coded as 0's*

*As we see from the heatmap the more sum of the TV show values are accumulated in the recent years and their dates are very close to today*

**So we can say Netflix has been focused on TV shows in recent years especially 2019...**
