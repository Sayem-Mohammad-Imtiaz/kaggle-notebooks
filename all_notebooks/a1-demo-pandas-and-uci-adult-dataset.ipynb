```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('../input/mlcourse/adult.data.csv')
data.head()
```

```python
data.info()
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
data['sex'].value_counts()
```

**2. What is the average age (*age* feature) of women?**

```python
data.loc[data.sex=="Female","age"].mean()
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
data.loc[data["native-country"]=="Germany","native-country"].count()/data["native-country"].count()
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year? **

```python
# You code here
data.groupby("salary")["age"].agg([np.mean,np.std])
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters* or *Doctorate* feature)**

```python
data[data["salary"]==">50K"].groupby("education")["education"].count()
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
# Displaying age statistics for each race and gender
data.groupby(["race","sex"])["age"].describe()
```

```python
# Maximum age of men of Amer-Indian-Eskimo race.
data[(data["race"]=="Amer-Indian-Eskimo") & (data["sex"]=="Male")]["age"].agg(np.max)
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
male_cond = data["sex"]=="Male"
salary_cond = data["salary"]==">50K"
married_cond = (data["marital-status"].isin(["Married-AF-spouse","Married-civ-spouse","Married-spouse-absent"]))
```

```python
data.loc[male_cond & salary_cond & married_cond,"marital-status"].count()
```

```python
data.loc[male_cond & salary_cond & ~(married_cond),"marital-status"].count()
```

```python
# QA
data.loc[(male_cond & salary_cond)].groupby("marital-status")["marital-status"].count()
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
max_hours = data["hours-per-week"].max()
print(f"Maximum number of hours: {max_hours}")

max_hours_cond = data["hours-per-week"]==max_hours
num_people_max_hours = data[max_hours_cond]["hours-per-week"].count()
print(f"Number of people who work maximum number of hours: {num_people_max_hours}")

salary_cond = data["salary"]==">50K"
num_people_max_hours_sal = data[(max_hours_cond) & (salary_cond)]["hours-per-week"].count()
# writing style and round function
print(f"% of people who work maximum number of hours and earn a lot: { \
                                round((num_people_max_hours_sal/num_people_max_hours)*100,2)} %")
```

**10. Count the average time of work (*hours-per-week*) for those who earn a little and a lot (*salary*) for each country (*native-country*). What will these be for Japan?**

```python
pd.crosstab(data["salary"],data["native-country"],values=data["hours-per-week"],aggfunc=np.mean)
```

```python
data[data["native-country"]=="Japan"].groupby(["native-country","salary"])["hours-per-week"].agg(np.mean)
```
