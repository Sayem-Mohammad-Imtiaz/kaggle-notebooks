```python
import numpy as np
import pandas as pd
import tensorflow
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
data=pd.read_csv("../input/pima-indians-diabetes-database/diabetes.csv")
data.head()
```

```python
x=data.drop(['Outcome'],axis=1)
y=data['Outcome']
```

```python
from sklearn.preprocessing import StandardScaler
Sc=StandardScaler()
x=Sc.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
model=Sequential()
model.add(Dense(32,input_dim=8,activation='sigmoid'))
model.add(Dense(18,activation='tanh'))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,epochs=100,verbose=2,batch_size=20)
```

```python
y_pred=model.predict_classes(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
a = accuracy_score(y_pred,y_test)
print('Accuracy is:', a*100)
```

```python
history = model.fit(x_train, y_train,validation_data = (x_test,y_test), epochs=100,verbose=2)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss']) 
plt.plot(history.history['val_loss']) 
plt.title('Model loss') 
plt.ylabel('Loss') 
plt.xlabel('Epoch') 
plt.legend(['Train', 'Test'], loc='upper left') 
plt.show()
```
