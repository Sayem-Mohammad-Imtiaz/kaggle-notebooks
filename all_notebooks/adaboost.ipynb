```python
# importing the packages
from numpy import mean
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score
import numpy as np 
import pandas as pd 
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        locat=os.path.join(dirname, filename)
        print(locat)
```

```python
# importing the dataset
dataset = pd.read_csv(locat)
# taking the values from the dataframe
data1 = dataset.values
# seperating the inputs and the output 
x = data1[:,:-1]
y = data1[:,-1]
```

```python
# check the type of the inputs
dataset.dtypes
# after examining the types in the datafram, found that amonia column is string and contain '#NUM!'
# thus first let's remove the row that has '#NUM!'
f = np.where(x[:,1]=='#NUM!')
s = f[0][:]
x = np.delete(x,s,0)
y = np.delete(y,s,0)
print(s)
# Now let's convert string to float
x[:,1] = x[:,1].astype(np.float)
print(x)
```

```python
# the model
model = AdaBoostClassifier()
cv = RepeatedStratifiedKFold(n_splits=15,n_repeats=10, random_state=6)
n_score = cross_val_score(model,x,y,cv=cv, scoring='accuracy', n_jobs=-1)
print('Accuracy = %0.2f percent' %(mean(n_score)*100))
```

```python
# Since it has a good accuracy let's use the model to predict 
# will use all the data for training except the first row becuase it will be used for prediction
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y.astype(np.float),test_size=0.002)
model.fit(x_train,y_train)
table = pd.DataFrame(columns=['Prediction','Actual value', 'Similar'])
for i in range(0,len(y_test)):
    t = model.predict(np.reshape(x_test[i,:],(1,20)))
    n_true ="False"
    if t==y_test[i]:
        n_true = 'True'
    table.loc[i] = [t,y_test[i],n_true]
print(table)
```
