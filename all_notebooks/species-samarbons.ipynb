```python
### Reading csv data

data = pd.read_csv("../input/iris-flower-dataset/Iris.csv")
```

```python
## Checking data
data
```

```python
### Processing raw data

data = data.drop(['Id'],axis = 1)
```

```python
### Null Check

data.isnull().any()
```

```python
### Check Data

data
```

```python
### Importing data splitting model

from sklearn.model_selection import train_test_split
```

```python
### Assigning target and feature variables

Y = data['Species']
X = data.drop(['Species'], axis = 1)
```

```python
### Splitting data into training/testing 70:30

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.3, random_state = 2)
```

```python
### checking set sizes
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python

Y_test.shape
```

```python
### importing KNN algorithm

from sklearn.neighbors import KNeighborsClassifier
```

```python
### Defining the model
### in this case k is 6

model = KNeighborsClassifier(n_neighbors = 6)

### Training the model

model.fit(X_train, Y_train)
```

```python
### Generating predictions from the model on the test set

predictions = model.predict(X_test)
```

```python
### Checking predictions

print(predictions)
```

```python
### importing accuracy_score and confusion_matrix 

from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
### Testing accuracy between actual values and predictions
### For k = 1 through 6, accuracy_score = 1.0 (Accuracy 100% )

accuracy_score(Y_test, predictions)
```

```python
### Confusion matrix 

confusion_matrix(Y_test, predictions)
```
