### How I almost got the perfect dataset

I took part in the **Getting Started with NLP** [competition](https://www.kaggle.com/c/nlp-getting-started) to understand a little more about the habits of a good data scientist. The community is extremely helpful. The following code was borrowed from user `@szelee` who highlighted that the competition main objective was already known. And so he provided another dataset that classified tweets according to their relevance to disaster. Anyway... needless to say that it explains why the first 400 submissions all hit 100%.

```python
import pandas as pd

test_df = pd.read_csv('/kaggle/input/nlp-getting-started/test.csv')
gt_df = pd.read_csv('/kaggle/input/disasters-on-social-media/socialmedia-disaster-tweets-DFE.csv')

gt_df = gt_df[['choose_one', 'text']]
gt_df['target'] = (gt_df['choose_one'] == 'Relevant').astype(int)
gt_df['id'] = gt_df.index
merged_df = pd.merge(test_df, gt_df, on='id')
sub_df = merged_df[['id', 'target']]

sub_df.to_csv('submission.csv', index=False)
```
