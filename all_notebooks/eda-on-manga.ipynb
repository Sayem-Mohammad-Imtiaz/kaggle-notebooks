```python
df.drop(14825,inplace=True)
```

```python
df['Rating'] = df['Rating'].apply(float)
```

```python
rng = df.shape[0]
```

```python
for i in range(rng):
    if i == 14825:
        continue
    val = df.loc[i,'Rating']
#     print(val)
    if val < 0 or val > 5:
        df.loc[i, 'Rating'] = pd.NA
```

```python
df['Rating'] = df['Rating'].ffill()
```

```python
df['Rating_Range'] = pd.cut(df['Rating'],bins=[0,1.5,3,4.5,5])
```

```python
df['Rating_Range'].value_counts().plot(kind="bar")
```

```python
vals = df['Latest Chapter'].str.startswith('Chapter')
```

```python
indexes = np.where(df['Latest Chapter'].str.startswith('Chapter'))[0]
```

```python
chaps = df[vals]['Latest Chapter'].str.split(' ')
```

```python
nums = []
```

```python
for ch in chaps:
    for chp in ch:
        if chp.isdigit():
            try:
                nums.append(int(chp))
            except:
                continue
            break
```

```python
chaps_num = pd.DataFrame(nums,columns=['Chapter Number'])
```

```python
plots = chaps_num['Chapter Number'].value_counts()[:20]
```

```python
plt.ylabel("Number of Mangas with Chapters Number on x-axis")
plt.xlabel("Chapter Number")
plt.title("Mangas and Chapter Number")
plt.bar(plots.index,plots);
```
