```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbn
import missingno as msno
from warnings import filterwarnings
filterwarnings('ignore')
veri = pd.read_csv("../input/individuals-killed-by-the-police/Police Fatalities.csv")
veri.sort_values("Date")
veri.info()
veri
```

```python
veri["Age"] = veri["Age"].fillna("33")
```

```python
veri["Gender"] = veri["Gender"].fillna("Male")
```

```python
veri["Race"] = veri["Race"].fillna("White")
```

```python
veri["Race"] = veri["Race"].fillna("White")
```

```python
veri["City"] = veri["City"].fillna("Los Angeles")
```

```python
veri.isnull().sum()
```

```python
veri["Age"] = veri["Age"].astype("int32")
veri["Gender"] = veri["Gender"].astype("category")
veri["Race"] = veri["Race"].astype("category")
veri["City"] = veri["City"].astype("str")
```

```python
veri.info()
```

```python
fig,ax = plt.subplots(7,1,figsize=(30,20))
sbn.countplot(veri["Race"],ax=ax[0]);
sbn.countplot(veri["Gender"],ax=ax[1]);
sbn.countplot(veri["Flee"],ax=ax[2]);
sbn.countplot(veri["State"],ax=ax[3]);
sbn.countplot(veri["Mental_illness"],ax=ax[4]);
sbn.countplot(veri["Armed"],ax=ax[5]);
sbn.countplot(veri["Age"],ax=ax[6]);
plt.xticks(rotation=45);
sbn.despine(bottom=True,left=True);
```

```python
fig,ax = plt.subplots(5,1,figsize=(30,20))
sbn.countplot(veri["Race"],ax=ax[0],hue=veri["Gender"]);
sbn.countplot(veri["Gender"],ax=ax[1],hue=veri["Flee"]);
sbn.countplot(veri["Flee"],ax=ax[2],hue=veri["State"]);
sbn.countplot(veri["State"],ax=ax[3],hue=veri["Mental_illness"]);
sbn.countplot(veri["Age"],ax=ax[4],hue=veri["Race"]);
plt.xticks(rotation=45);
sbn.despine(bottom=True,left=True);
```
