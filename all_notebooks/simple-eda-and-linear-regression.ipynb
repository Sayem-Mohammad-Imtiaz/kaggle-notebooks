# Basic EDA and Linear Regression

In this notebook, we'll explore the dataset and run a simple linear.

As we can see, the main info are the area, no. of rooms, longitude, and latitude.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import rcParams
```

```python
plt.rcParams['figure.figsize'] = (10,8)
sns.set_theme(style="darkgrid")
```

```python
df = pd.read_csv('../input/amsterdam-house-price-prediction/HousingPrices-Amsterdam-August-2021.csv')
df
```

Removing the columns that we'll not use and then remove the null rows.

```python
df = df [['Price','Area','Room','Lon','Lat']]
df['Price'] = df['Price'].astype(float)
```

```python
df.isnull().values.any()
```

```python
df = df.dropna()
```

Now we'll plot some graphs to see how the data is distributed, and how the features are related.

```python
sns.relplot(x="Area", y="Price", hue="Room", data=df)
plt.show()
```

```python
sns.pairplot(df)
```

```python
sns.heatmap(df.corr(), annot=True, fmt=".2f")
plt.show()
```

# Linear regression
Now we'll apply a linear regression. First, we'll divide the dataset into train and test.

```python
from sklearn import  linear_model
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import xgboost
```

```python
X = df[['Area', 'Room', 'Lat','Lon']]
y = df[['Price']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

Now we'll fit the data and see the results.

```python
regression = linear_model.LinearRegression()
regression.fit(X_train, y_train)
```

```python
y_pred = regression.predict(X_test)
print('Coefficient of determination: %.2f' % r2_score(y_test, y_pred))
```

There is room for improvement, for a more robust model.
