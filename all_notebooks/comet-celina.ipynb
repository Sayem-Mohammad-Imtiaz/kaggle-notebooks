The data set focuses on the recommended anime by 73,516 users from myanimelist.net. 

Questions:
1. What are the common factors of the top 20 anime (default list)?
2. What are the top 20 anime that has the most number of members?
3. What is the relation of the rating of an anime and its number of members
4. What is the most common genre in anime?

    There are many approaches in being able to find the best anime. One approach it to look at its ranking. Another approach is to look at its number of followers. This report would focus on these 2 approaches. The goal of this report is to determine the difference and similarity that 2 approaches may have.  

    There are many factors that makes an anime popular. From the data set provided, majority of the top 20 anime were TV shows. The average rating of the animes is 9.0635. The data also shows that an anime that has a shounen genre is more likely to be popular. Half of the top 20 anime also had drama as their genre. There is a weak relationship between an anime's number of episodes and number of members. The top 20 anime also shows that there is no relation with an anime's rating and its number of members. 

    When the data is arranged by the number of members, the average rating of the animes is 8.3615. All 20 anime were TV shows. The table showed that out of 20 anime, 15 had action as one of their genre. The genre, supernatural has the second most frequency. There is a weak relationship between an anime's number of episodes and number of members. Unlike the first approach, this approach shows that there is a strong relationship with an anime's rating and its number of members.

    The first approach shows that there is no relation with an anime's rating and its number of members. The second approach, on the other hand, shows that there is a strong relation. When all of the anime in data set were used, it showed that there is a strong relation between the 2 variables. This suggests that having more members mean that an anime will have a higher rating.

    The 2 approaches their top genre of anime differs. With 10 occurrences, the first approach shows that the shounen is the top genre. meanwhile, The second approach had 9 occurrences for shounen. The second approach had 15 counts for action while the first approach had 9. While the top genres are different, the genre that the anime of the approaches have are almost the same. 

    The data from the 2 approaches suggests that animes that are TV shows tend to be more recommended. It also shows that the genres that does the best are action, shounen, and drama. Users can look at the type and genre of an anime to look for a recommendation. To narrow down their search, they can look at the most popular animes. Animes also tend to have higher ratings when there are more members. This suggests that it would also be beneficial to look at the number of members an anime has when looking for a recommendation.

```python
e = top20['episodes']
r = top20['rating']
e = pd.to_numeric(e, errors='coerce')
r = pd.to_numeric(r, errors='coerce')

print(np.mean(e))
print(np.max(e))
print(np.min(e))

df = pd.DataFrame({'rating': r, 'episodes': e})
df.corr('pearson')
```
