```python
basic_income_info = pd.read_csv("../input/basic_income_dataset_dalia.csv")
df = pd.DataFrame(basic_income_info)
corr = df.corr()
corr
```

Distribution of respondents by age and gender

```python
ages = df.age.unique()
counts = df.groupby(['age','gender']).count()
totals = counts.sum(level=0)
counts = counts.unstack(level=1)
counts.columns = counts.columns.droplevel(level=0)
counts=counts.ix[:,0:2]
plt.bar(ages, counts['male'], bottom=None, color='blue', label='M')
plt.bar(ages, counts['female'], bottom=counts['male'], color='pink', label='F')
plt.legend()
plt.xlabel('Ages')
plt.ylabel('Count')
genders = df.groupby(['gender']).count()
genders=genders.ix[:,0:1]
a = plt.axes([0.95, 0.45, .3, .45], facecolor='y')
plt.pie([genders.country_code['male'],genders.country_code["female"]],colors=['blue','pink'])
```
