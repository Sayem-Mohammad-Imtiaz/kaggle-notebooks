```python
songs_mat = songs[['Beats.Per.Minute', 'Energy',
       'Danceability', 'Loudness..dB..', 'Liveness', 'Valence.', 'Length.',
       'Acousticness..', 'Speechiness.', 'Popularity']]
```

```python
plt.figure(figsize=(14,8))
sns.heatmap(songs_mat.corr(),annot=True)
```
