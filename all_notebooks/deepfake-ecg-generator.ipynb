```python
import os
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
import ecg_plot
```

```python
os.mkdir("electrocardiograms")
```

```python
deepfakeecg.generate(50, "/kaggle/working/electrocardiograms/",start_id=0, run_device="cpu",) # Generate 5 ECGs to the current folder starting from id=0
```

```python
j = 0
for i in os.listdir("/kaggle/working/electrocardiograms/"):
    j = j + 1
    ecg = np.zeros([5000,12])
    ecg_temp = np.loadtxt("/kaggle/working/electrocardiograms/"+i)
    ecg[:,0:2] = ecg_temp[:,0:2]
    ecg[:,2]= ecg[:,1]- ecg[:,0]
    ecg[:,3] = -0.5 * (ecg[:,0] + ecg[:,1])
    ecg[:,4] = ecg[:,0]-0.5 * ecg[:,1]
    ecg[:,5] = ecg[:,1]-0.5 * ecg[:,0]
    ecg[:,6:] = ecg_temp[:,2:]
    plt.figure(figsize=(20,8))
    ecg_plot.plot_12(ecg.T*(2/1000), title="Deep Fake ECG {}".format(j))
    plt.show()
```
