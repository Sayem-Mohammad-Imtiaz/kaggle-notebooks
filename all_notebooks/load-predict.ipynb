```python
# IMPORT REQUIRED LIBRARIES
import numpy as np
import pandas as pd
%matplotlib inline

# IMPORT LIBRARIES FOR VISUALIZATION
import matplotlib.pyplot as plt


import sklearn
from sklearn.linear_model import LogisticRegression 
```

```python
# READ DATA FILE
df = pd.read_csv('../input/loan-train/loan_train.csv')
df.describe()
```

```python
loanstat_map = {'Y':1, 'N':0}
df['Loan_Status'] = df['Loan_Status'].map(loanstat_map)

edu_map = {'Graduate':1, 'Not Graduate':0}
df['Education'] = df['Education'].map(edu_map)

area_map = {'Urban':3, 'Semiurban':2, 'Rural':1}
df['Property_Area'] = df['Property_Area'].map(area_map)

df = df.fillna({'Gender':'Male'})
sex_map = {'Male':1, 'Female':0}
df['Gender'] = df['Gender'].map(sex_map)

df = df.fillna({'Married':'Yes'})
wed_map = {'Yes':1, 'No':0}
df['Married'] = df['Married'].map(wed_map)

df = df.fillna({'Self_Employed':'No'})
self_empl_map = {'Yes':1, 'No':0}
df['Self_Employed'] = df['Self_Employed'].map(self_empl_map)

df = df.fillna({'Credit_History':0})

LoanAmount_mean = df['LoanAmount'].mean()
df = df.fillna({'LoanAmount':LoanAmount_mean})

Loan_Amount_Term_mean = df['Loan_Amount_Term'].mean()
df = df.fillna({'Loan_Amount_Term':Loan_Amount_Term_mean})

df.head()
```

```python
m = np.random.rand(len(df)) < 0.9

train = df[m]
test = df[~m]

#train_x = train[['Gender', 'Married', 'Education', 'Self_Employed', 'ApplicantIncome', 'Loan_Amount_Term', 'LoanAmount', 'Property_Area', 'Credit_History']]
train_x = train[['Education', 'Self_Employed', 'ApplicantIncome', 'Loan_Amount_Term', 'LoanAmount', 'Property_Area', 'Credit_History']]

train_y = train['Loan_Status']

#test_x = test[['Gender', 'Married', 'Education', 'Self_Employed',  'ApplicantIncome','Loan_Amount_Term', 'LoanAmount', 'Property_Area', 'Credit_History']]
test_x = test[['Education', 'Self_Employed',  'ApplicantIncome','Loan_Amount_Term', 'LoanAmount', 'Property_Area', 'Credit_History']]

test_y = test['Loan_Status']
```

```python
# IMPORT LOGISTIC REGRESSION MODEL
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

LR = LogisticRegression(C=0.01, solver='liblinear').fit(train_x, train_y)
LR
```

```python
test_y_pred = LR.predict(test_x)     # LETS CHECK THE PREDICTED VALUES FOR LOAN STATUS
test_y_pred[0:5]
```

```python
test_y_pred_prob = LR.predict_proba(test_x)
test_y_pred_prob[0:5]
```

```python
print("Mean absolute error: %.5f" % np.mean(np.absolute(test_y_pred - test_y)))
```

```python
from sklearn.metrics import classification_report

print(classification_report(test_y, test_y_pred))
```
