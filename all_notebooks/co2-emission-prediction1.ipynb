```python
data1=df[df['Engine Size(L)']<df['Engine Size(L)'].quantile(0.975)]
data1=data1[data1['CO2 Emissions(g/km)']<data1['CO2 Emissions(g/km)'].quantile(0.98)]
```

```python
plt.boxplot(data1['Engine Size(L)'])
```

```python
plt.scatter(data1['Engine Size(L)'],data1['CO2 Emissions(g/km)'])
```

```python
plt.boxplot(data1['CO2 Emissions(g/km)'])
```

```python

X=data1['Engine Size(L)']
Y=data1['CO2 Emissions(g/km)']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=0.8)
#spliting data into 80% train,20% test,0% validation
```

```python
from sklearn.linear_model import LinearRegression#importing LinearRegression class

Engine_S = LinearRegression()#object creation
X_train = np.array(X_train).reshape(-1,1)
X_test = np.array(X_test).reshape(-1,1)
Engine_S.fit(X_train, y_train)
```

```python
E=Engine_S.coef_[0]
E1=Engine_S.intercept_
print("E = ",E,'\nE1 = ',E1)
```

```python
y_pred=(X)*E+E1#using values of slope and intercept
#print(y_pred.head(5))
plt.title('Co2 Emission/Engine Size')
plt.scatter(X,Y)
plt.scatter(X,y_pred,marker='s')
plt.xlabel('Engine Size(L)')
plt.ylabel('CO2 Emissions(g/km)')
```

```python
y_pred=(X_test)*E+E1#using values of slope and intercept
#print(y_pred.head(5))
plt.scatter(X,Y)
plt.scatter(X_test,y_pred,color='black')
```

```python
y_pred=Engine_S.predict(X_test)
plt.scatter(X,Y)
plt.scatter(X_test,y_pred)#for test values
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print('Mean absolute error :',mean_absolute_error(y_test,y_pred))#mean absolute error for orginal test data and predicted test data
print('Mean squared error :',np.sqrt(mean_squared_error(y_test,y_pred)))#mean squared error for orginal test data and predicted test data
```
