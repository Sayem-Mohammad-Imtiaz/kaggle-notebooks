```python
#Line Plot
#grafik tabloları
#data2.NA_Sales.plot(kind = 'line',color='b', label='USA_Satış',linewidth=1, alpha=1, grid=True, linestyle=':')
data1[data1['Rank']<200].NA_Sales.plot(kind = 'line', color = 'r',label = 'USA_Satış',linewidth=1,alpha = 1,grid = True,linestyle = ':')
#data1. verinin geldiği, [data1['Rank']<200] filtrelediğin yer NA_Sales. listelemek istediğin kolon plot listeme fonksiyonu
#kind tablo türünün belirtildiği bölüm,label değere vermek istediğin isim, alpha saydamlık, Grid arkadaki çizgiler, linestyle grafik çizim türü
plt.legend(loc='upper left')     # legend = puts label into plot grafik ismi
#(loc='upper left') grafik isim konumu
plt.xlabel('Rank')              # label = name of label
plt.ylabel('Satış sayısı')
plt.title('Grafik başlığı')            # title = title of plot
plt.show()
```
