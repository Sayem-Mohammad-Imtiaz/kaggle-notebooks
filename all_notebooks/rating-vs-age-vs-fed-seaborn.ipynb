<h3>Importing relevant libraries</h3>

```python
import seaborn as sns
```

```python
data = pd.read_csv('/kaggle/input/top-women-chess-players/top_women_chess_players_aug_2020.csv')
```

```python
data.head()
```

<h3>Rating vs title

```python
sns.pairplot(data[['Title','Standard_Rating','Rapid_rating','Blitz_rating']],height=5,hue='Title')
```

Observation:
1. Rating: GM>IM>WGM>WIM>WFM>WCM
2. Ratings increase linearly across game formats
3. Improvement in Blitz rating by one unit results in Rapid and Standard rating to improve by more than one unit

<h3>Age vs title

```python
sns.catplot(x="Title", y="Year_of_birth", jitter=False, data=data, kind='violin')
```

Observation:
1. GM,IM,WGM,WIM peak at 30-35 years of age
2. Title holders younger than 20 years of age are outliers
3. Most probable age for title holders is 20-40 years

<h3>Federation vs title

```python
data_2=data[(data['Title']=='GM')|(data['Title']=='IM')]
sns.catplot(x="Federation", y="Year_of_birth", data=data_2, row='Title',kind='swarm',height=4, aspect=4);
```

Observation:
1. Recent GMs are from only select few countries including HUN,CHN,IND,RUS
2. Recent IMs are from a different set of countries including USA,QAT,BUL,KAZ
3. There are ~15 countries with IMs but no GMs
