# How will I figure it out?

I will first do a graph of the general trend of the attribute. Secondly to figure out which one is most standard deviations above the **previous** year

# What attributes will I assess?

transistors (millions), memory size (megabytes), memory speed (gb/sec) and GFLOPS (float32)

scroll down to see what the most influential GPUs are in each of the 4 metrics

imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

get data

```python
dataset = pd.read_csv("../input/amd-radeon-and-nvidia-gpu-specifications/gpus_v2.csv")
```

define functions

```python
def plot(years, ys, feature=''):
    dict = {}

    for i in range(min(years), max(years) + 1):
        dict[i] = []
    for i in range(len(ys)):
        dict[years[i]].append(ys[i])
    for i in dict:
        dict[i] = np.mean(dict[i])
    
    y=[]
    year=[]
    for i in dict:
        year.append(i)
        y.append(dict[i])
        
    plt.plot(year, y)
    plt.ylabel(feature)
    plt.xlabel('year')
    plt.show()
    
    
def probs(years, ys):
    dict = {}

    for i in range(min(years), max(years) + 1):
        dict[i] = []
    for i in range(len(ys)):
        dict[years[i]].append(ys[i])
    for i in dict:
        mean = np.mean(dict[i])
        stddev = np.std(dict[i])
        dict[i]=[mean, stddev]
        
    return dict


def best(years, ys, names):
    probabilities = []
    name_list = []
    info = probs(years, ys)
    for i in range(len(ys)):
        if years[i] != min(years):
            if info[years[i]-1][0] < ys[i]:
                probabilities.append(norm.pdf(
                    ys[i], loc=info[years[i]-1][0], scale=info[years[i]-1][1]))
                name_list.append(names[i])
    best = min(probabilities)
    return name_list[probabilities.index(best)]
```

# GFLOPS - Radeon HD 4870 X2

```python
plot(dataset["Year"], dataset["GFLOPS"], 'Memory speed')
```

```python
print(best(dataset["Year"], dataset["GFLOPS"],dataset["Name"]))
```

# Transistor count - GRID A100

```python
plot(dataset["Year"], dataset["Transistors (mln)"], 'Memory speed')
```

```python
print(best(dataset["Year"], dataset["Transistors (mln)"],dataset["Name"]))
```
