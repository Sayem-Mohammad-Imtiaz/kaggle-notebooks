```python
data = pd.read_csv('/kaggle/input/diabetes-dataset/diabetes2.csv')
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(data.drop(['Outcome'],axis=1),data.Outcome,random_state=42,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(C=2.5,max_iter=100).fit(X_train,y_train)
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_train,model.predict(X_train)))
```

```python
confusion_matrix(y_train,model.predict(X_train))
```

```python
confusion_matrix(y_test,model.predict(X_test))
```
