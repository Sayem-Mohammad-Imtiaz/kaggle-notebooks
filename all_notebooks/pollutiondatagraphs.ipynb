```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pollution = pd.read_csv("../input/top100polutionmay2021/May2021Top100Polution.csv")
```

```python
pollution.describe()
```

```python
pollution.head()
```

```python
import plotly.express as px
fig = px.bar(pollution, y='Pollution Index', x='Country')
fig.show();
```

```python
import plotly.express as px
fig = px.bar(pollution, y='Exp Pollution Index', x='Country')
fig.show();
```

```python
import plotly.express as px
fig = px.pie(pollution, values='Pollution Index', names='Country')
fig.update_traces(textposition='inside')
fig.update_layout(uniformtext_minsize=12, uniformtext_mode='hide')
fig.show();
```

```python
import plotly.express as px
fig = px.histogram(pollution, x="Country", color="Pollution Index", marginal="box")
fig.show();
```

```python
fig = px.scatter(
    pollution, x='Pollution Index', y='Rank', opacity=0.65,
    trendline='ols', trendline_color_override='darkblue'
)
fig.show();
```

```python
import seaborn as sns
sns.pairplot(data=pollution);
```

```python
sns.jointplot(data=pollution, x="Rank", y="Pollution Index");
```

```python
sns.jointplot(data=pollution, x="Rank", y= "Exp Pollution Index");
```
