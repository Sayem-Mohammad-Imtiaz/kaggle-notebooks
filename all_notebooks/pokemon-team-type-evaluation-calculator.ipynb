## Pokemon Team Type Evaluation Calculator ##

This is a quick little calculator to help evaluate a team of 6 pokemon in terms of their type coverage. It looks at the types of the 6 pokemon you have and compares them collectively against all of the types in the game to see what you'll be best and worst against. It assumes that each pokemon is attacking with the same move types as their own type, but of course its possible to train other moves to help cover your weak points.


To use this notebook, simply fork it, edit the variables at the top of the code below to choose your pokemon and then publish!
