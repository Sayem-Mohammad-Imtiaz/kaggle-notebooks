Observe thar, min and max value for each column are finite. This ensures that the dataset doesn't contain any infinite values.
We have the mean of the values for each column. So now we have to use fillna() to fill up those missing values

```python
dataset.age = dataset.age.fillna(dataset.age.mean())
dataset.height = dataset.height.fillna(165.805794)
```
