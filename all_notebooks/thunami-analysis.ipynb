```python
sou=pd.read_csv("../input/sources.csv")
```

```python
sou.head(2)
```

```python
sou.shape
```

```python
from ggplot import *
ggplot(aes(x='LATITUDE', y='LONGITUDE', color='DAMAGE_TOTAL'), data=sou) + \
    geom_point() + \
    scale_color_gradient(low = 'red', high = 'blue')
```

```python
sou.isnull().sum().sort_values().plot(kind='barh',
                                             legend=False,
                                             figsize=(15,20),
                                             title="Null Count",
                                             fontsize=15,
                                             alpha=0.5)
```

```python
#Group by Countries
sou.groupby(sou.COUNTRY)['COUNTRY'].count().sort_values(ascending=False).head(15).plot(kind='pie',
                                             legend=False,
                                             figsize=(15,10),
                                             title="Country wise count",
                                             fontsize=15)
```

```python
sou.groupby(sou.YEAR)['YEAR'].count().sort_values(ascending=False).head(15).plot(kind='pie',
                                             legend=False,
                                             figsize=(15,10),
                                             title="Year wise Count",
                                             autopct='%.1f%%',
                                             fontsize=15)
```

```python
sou.groupby(sou.YEAR)['YEAR'].count().sort_values(ascending=False).head(15).plot(kind='bar',
                                             legend=False,
                                             figsize=(15,10),
                                             title="Year wise Count",
                                             fontsize=15)
```

```python
plt.figure(figsize=(20,5))
sns.pointplot(x='PRIMARY_MAGNITUDE',y='CAUSE',data=sou)
```
