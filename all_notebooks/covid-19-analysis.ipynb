```python
data = pd.read_csv("/kaggle/input/who-covid19-data-tabe/WHO COVID-19 global table data August 11th 2021 at 10.41.34 AM.csv")
```

```python
data.head()
```

```python
data=data.reset_index()
```

```python
data.rename(columns =  {'Name': 'Continent'}, inplace = True)
data.rename(columns =  {'index': 'Country'}, inplace = True) 
```

```python
data.rename(columns =  {'Name': 'Continent'}, inplace = True)
data.rename(columns =  {'index': 'Country'}, inplace = True)  
data.rename(columns =  {'WHO Region': 'Cases - cumulative total real'}, inplace = True)
data.rename(columns =  {'Cases - newly reported in last 7 days per 100000 population': 'Cases - newly reported in last 24 hours real'}, inplace = True) 
data.rename(columns =  {'Cases - newly reported in last 24 hours': 'Deaths - cumulative total real'}, inplace = True) 
data.rename(columns =  {'Deaths - cumulative total': 'df'}, inplace = True) 
```

```python
data['Country']= data['Country'].replace('United States of America', 'United States')
data['Country']= data['Country'].replace('Russian Federation', 'Russian')
data['Country']= data['Country'].replace('The United Kingdom' , 'United Kingdom')
data['Country']= data['Country'].replace('Iran (Islamic Republic of)' , 'Iran')
data['Country']= data['Country'].replace('Czechia' , 'Czech Republic')
data['Country']= data['Country'].replace('occupied Palestinian territory, including east Jerusalem' , 'Palestine, State of')
data['Country']= data['Country'].replace('Republic of Moldova' , 'Moldova')
data['Country']= data['Country'].replace('Viet Nam' , 'Vietnam')
data['Country']= data['Country'].replace('Bolivia (Plurinational State of)' , 'Bolivia')
data['Country']= data['Country'].replace( 'Palestine','Palestine, State of')
data['Country']= data['Country'].replace('Venezuela (Bolivarian Republic of)' , 'Venezuela')
data['Country']= data['Country'].replace('Republic of Korea' , 'Korea, Republic of')
data['Country']= data['Country'].replace( 'Korea(Republic of','Korea, Republic of')
data['Country']= data['Country'].replace( 'Bosnia and Herzegovina','Bosnia And Herzegovina')
data['Country']= data['Country'].replace( ' Namibia ','Namibia')
data['Country']= data['Country'].replace( 'Syrian Arab Republic','Syria')
data['Country']= data['Country'].replace( 'United Republic of Tanzania','Tanzania')
```

```python
data = data[0:][1:]
```

```python
GroupedByContinents = data.groupby('Continent').sum()
GroupedByContinents
```

```python
GroupedByContinents = GroupedByContinents.sort_values(by='Cases - cumulative total real')
```

```python
GroupedByContinents['Recovered'] = GroupedByContinents['Cases - cumulative total real'] - GroupedByContinents['Deaths - cumulative total real']
```

```python
GroupedByContinents
```

```python
barWidth = 0.25
plt.rcParams['figure.figsize'] = (20, 10)
# set heights of bars
bars1 = GroupedByContinents['Cases - cumulative total real']
bars2 = GroupedByContinents['Deaths - cumulative total real']
bars3 = GroupedByContinents['Recovered']
 
# Set position of bar on X axis
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
 
# Make the plot
plt.bar(r1, bars1, color='#7f6d5f', width=barWidth, edgecolor='white', label='Confirmed Cases')
plt.bar(r2, bars2, color='#557f2d', width=barWidth, edgecolor='white', label='Deaths')
plt.bar(r3, bars3, color='#2d7f5e', width=barWidth, edgecolor='white', label='Recovered')
 
# Add xticks on the middle of the group bars
plt.xlabel('Continents', fontweight='bold')
plt.ylabel('Count per Million', fontweight='bold')
plt.title('\nCovid 19 Cases/Death/Recovered by Continents \n', fontsize=26,fontweight='bold')
plt.xticks([r + barWidth for r in range(len(bars1))], ['Other','Western Pacific','Africa','East Mediterranean','South-East Asia','Europe','Americas'])
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
# Create legend & Show graphic
plt.legend()
plt.show()
```

```python
size = GroupedByContinents['Cases - cumulative total real']
colors = ['lightblue', 'lightgreen','purple','grey','orange','#2076a2','#f56642']
labels = "Others", "Western Pacific","Africa","East Mediterranean","South-East Asia","Europe","America"
explode = [0, 0.01,0.01,0.01,.01,.01,.01]

my_circle = plt.Circle((0, 0), 0.7, color = 'white')

plt.rcParams['figure.figsize'] = (10, 10)
plt.pie(size, colors = colors, labels = labels, shadow = True, explode = explode, autopct = '%.2f%%')
plt.title('Covid 19 cases per continent', fontsize = 20,fontweight='bold',color='red')
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.legend()
plt.show()
```
