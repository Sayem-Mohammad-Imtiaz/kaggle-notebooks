#### import pandas as pd

### Data aggregation and time-series preprocessing

Simulated labeled text

```python
data_path = '../input/texts-labeled-timestamp/simulated_texts_labeled.csv'
```

```python
df = pd.read_csv(data_path, index_col=0)
```

```python
df.head()
```

```python
df[df.label=='A']
```

```python
# Let's make sure we don't have any duplicated rows
```

```python
df = df.drop_duplicates()
```

```python
# Set timestamp index
```

```python
df_index = df.set_index(pd.to_datetime(df.timestamp))
```

### We want to count the different text messages by category for a given time interval

```python
# Let's groupby first by label and then resample for every 5 minutes - counting occurrencies-

df_aggregated = df_index.groupby(['label']).resample('5T').count()
```

```python
df_aggregated.head()
```

```python
# Not exactly descriptive names, let's use more descriptive names and de-duplicate the timestamp column
```

```python
df_aggregated=df_aggregated.rename(columns={'label':'count', 'timestamp':'timestamp_'})
```

```python
df_aggregated.head()
```

```python
# Now let's flaten the dataframe by re-indexing, and get rid of the unnecessary columns
```

```python
df_aggregated = df_aggregated.reset_index()[['label','timestamp', 'count']]
```

```python
df_aggregated.sort_values(['label','timestamp'])
```

```python
# Split by label

series_A = df_aggregated[df_aggregated.label=='A'].set_index('timestamp')
series_B = df_aggregated[df_aggregated.label=='B'].set_index('timestamp')
series_C = df_aggregated[df_aggregated.label=='C'].set_index('timestamp')
```
