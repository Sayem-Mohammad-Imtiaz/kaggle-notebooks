```python
study_weight = [0.05, 0.4, 0.4, 0.15] # study1 0.6, study2 0.4
print('study-weight: ',study_weight)

def ensemble_study(row):
    test_type = row['id'].split('_')[1]
#     a_test['1'] = 1
    if test_type == 'image': return row

    test_name = row['id'].split('_')[0]
#     print(test_name)
    res1 = huang1[huang1['id'] == test_name+'_study'].values[0][2]
    res2 = huang2[huang2['id'] == test_name+'_study'].values[0][2]
    res3 = huang3[huang3['id'] == test_name+'_study'].values[0][2]
    res4 = huang4[huang4['id'] == test_name+'_study'].values[0][2]
#     print(res1)
    
    score1 = study_weight[0] * float(res1.split(' ')[1]) + study_weight[1] * float(res2.split(' ')[1]) + study_weight[2] * float(res3.split(' ')[1]) ++ study_weight[3] * float(res4.split(' ')[1])
    score2 = study_weight[0] * float(res1.split(' ')[7]) + study_weight[1] * float(res2.split(' ')[7]) + study_weight[2] * float(res3.split(' ')[7]) ++ study_weight[3] * float(res4.split(' ')[7])
    score3 = study_weight[0] * float(res1.split(' ')[13]) + study_weight[1] * float(res2.split(' ')[13]) + study_weight[2] * float(res3.split(' ')[13]) ++ study_weight[3] * float(res4.split(' ')[13])
    score4 = study_weight[0] * float(res1.split(' ')[19]) + study_weight[1] * float(res2.split(' ')[19]) + study_weight[2] * float(res3.split(' ')[19]) ++ study_weight[3] * float(res4.split(' ')[19])
    
    res = f'negative {score1} 0 0 1 1 typical {score2} 0 0 1 1 indeterminate {score3} 0 0 1 1 atypical {score4} 0 0 1 1'
    row['PredictionString'] = res
    return row

# huang = sample_sub.apply(lambda x:ensemble_study(x), axis = 1)


huang = huang1
huang.head(10)
```
