# **Importing data:**

```python
train_set=pd.read_csv("../input/linkedin-influencers-data/influencers_data.csv")
train_set.shape
train_set.head(2)
```

```python
import pandas_profiling
#descriptive statistics
pandas_profiling.ProfileReport(train_set)
```

From the above report, we can see that columns views and votes have significantly higher missing values. Let us drop them before proceeding further.

```python
train_set_new=train_set.drop(columns=['views','votes'])
train_set_new.columns
```

```python
train_set_new.shape
```

### **Top 10 people with highest number of followers having more than 500 connections:**

```python
train_set_new[train_set_new['connections']=='500+'].sort_values('followers',ascending=False)['name'].unique()[:10]
```

```python
#checking if a value exists in a column
print('India' in train_set_new['location'])
```

```python
train_set_new['location'].unique()
```

```python
train_set_new['location'].value_counts().plot(kind='bar')
```

###  **Top profiles with more than 50k followers from California state of United States**

```python
res=train_set_new[['name','headline','followers']].loc[(train_set_new['followers']>50000) & 
              train_set_new['location'].str.contains('California')].drop_duplicates().sort_values(by='followers',ascending=False)
print(res)
```

```python
train_set_new['time_spent'].head(5)
```

### **Top hashtags by number**

```python
train_set_new.nlargest(5,['num_hashtags'],keep='all')[['name','hashtags','num_hashtags']]
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (10, 5))
 
# creating the bar plot
plt.bar(train_set_new['comments'], train_set_new['reactions'], color ='maroon',
        width = 0.4)
 
plt.xlabel("Comments")
plt.ylabel("Reactions")
plt.title("Comments Vs reactions")
plt.show()
```
