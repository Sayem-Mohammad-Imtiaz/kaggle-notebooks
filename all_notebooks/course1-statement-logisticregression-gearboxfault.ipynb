```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

# Datasets import

```python
healthyDataset = pd.read_csv("/kaggle/input/gearbox-fault-diagnosis-elaborated-datasets/gearbox-fault-diagnosis-elaborated-datasets/stdev/healthy30hz_stdev_100.csv")
```

```python
brokenDataset = pd.read_csv("/kaggle/input/gearbox-fault-diagnosis-elaborated-datasets/gearbox-fault-diagnosis-elaborated-datasets/stdev/broken30hz_stdev_100.csv")
```

```python
dataset = pd.concat([healthyDataset, brokenDataset], axis=0 )
```

```python
dataset.describe()
```

# Predict failure using Logistic Regression

```python
# Create dataset of predictors, X
#columns = ['a1','a2','a3','a4','load']
columns = ['a1','a2','a3','a4']
X = dataset[columns]
```

```python
# Create dataset of predicted variable, y
columns = ['failure']
y = dataset[columns]
```

## [1] Separate dataset into training and test

```python
# Split the dataset: 80% train, 20% test
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split (X,y, test_size=0.2, random_state=0)
```

## [2] Train the model

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Setup the model
model = LogisticRegression(max_iter=1000)
```

```python
# Train the model
model.fit(X_train, y_train)
```

```python
# Predict test output
y_pred = model.predict(X_test)
print("Predict outcome for the first 10 ten rows:", y_pred[:10])
```

```python
print("Actual outcome for the first 10 ten rows:", y_test[:10])
```

### [2.1] Obtain intercept and coefficients

```python
model.intercept_
```

```python
model.coef_
```

```python
print("     stdev(a1),  stdev(a2),  stdev(a3),  stdev(a4),    load \n",
      "    ---------   ---------   ---------   ---------     ---- \n", model.coef_)
```

## [3] Evaluate the model

## [3.1] Confusion matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_pred = y_pred, y_true = y_test)
cm
```

## [3.2] Classification report

```python
from sklearn.metrics import classification_report
```

```python
report = classification_report(y_pred = y_pred, y_true = y_test)
print(report)
```
