```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('../input/us-graduate-schools-admission-parameters/US_graduate_schools_admission_parameters_dataset.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.drop('Serial No.', axis='columns')
```

```python
df.head()
```

```python
x = df.drop('Chance of Admit ', axis='columns')
y = df['Chance of Admit ']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=42)
```

```python
import xgboost as xgb
```

```python
my_model = xgb.XGBRegressor(learning_rate=0.01,
                            n_estimators=200,
                            max_depth = 10)
```

```python
my_model.fit(x_train,y_train)
```

```python
predictions = my_model.predict(x_test)
predictions
```

```python
score = my_model.score(x_test, y_test)
mae = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mae)
```

```python
print('Model Accuracy : ', score)
print('Mean Squared Error : ', mae)
print('RMSE : ', rmse)
```

```python
rmse
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'max_depth': range (2, 10, 1),
    'n_estimators': range(60, 220, 40),
    'learning_rate': [0.1, 0.01, 0.05]
}
```

```python
grid_search = GridSearchCV(estimator=my_model,
                           param_grid=params,                           
                           n_jobs=10,
                           cv=5,
                           verbose=True)
```

```python
grid_search.fit(x_train,y_train)
```

```python
predictions_1 = grid_search.predict(x_test)
predictions_1
```

```python
score_1 = grid_search.score(x_test, y_test)
mae_1 = mean_squared_error(y_test, predictions_1)
rmse_1 = np.sqrt(mae_1)
```

```python
score_1, mae_1, rmse_1
```

```python
print("Without GridsearchCV Results")
print('Model Accuracy : ', score)
print('Mean Squared Error : ', mae)
print('RMSE : ', rmse)
```

```python
print("GridsearchCV Results")
print('Model Accuracy : ', score_1)
print('Mean Squared Error : ', mae_1)
print('RMSE : ', rmse_1)
```
