```python
df_train = pd.read_csv('../input/mobile-price-classification/train.csv')
df_test = pd.read_csv('../input/mobile-price-classification/test.csv')
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_train.loc[:,df_train.columns != 'price_range'],
                                                    df_train['price_range'],
                                                    test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

forest_model = RandomForestClassifier(n_estimators=700,random_state=1)
forest_model.fit(x_train, y_train)
melb_preds = forest_model.predict(x_test)
print(accuracy_score(y_test, melb_preds))
```
