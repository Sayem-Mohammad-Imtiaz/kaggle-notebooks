# **Exploring the data**

```python
df = pd.read_csv("/kaggle/input/3wine-classification-dataset/wine.csv")
df
```

```python
df["Wine"].unique()
```

# **Store target as y**

```python
y = df[["Wine"]].copy()
y
```

# **Store features as X**

```python
df.columns
```

```python
features = ['Alcohol', 'Malic.acid', 'Ash', 'Acl', 'Mg', 'Phenols',
       'Flavanoids', 'Nonflavanoid.phenols', 'Proanth', 'Color.int', 'Hue',
       'OD', 'Proline']
X = df[features].copy()
X
```

# **Train, test and split then fit data**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 324)
```

```python
clf = DecisionTreeClassifier(max_leaf_nodes = 5, random_state = 100)
clf.fit(X_train, y_train)
```

# **Predict wine type(1, 2, or 3)**

```python
predictions = clf.predict(X_test)
predictions[:15]
```

```python
y_test[:15]
```

# **Measure accuracy of classifier = 89.8%**

```python
accuracy_score(y_test, predictions)
```
