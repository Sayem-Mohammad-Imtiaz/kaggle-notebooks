```python
newdata=df.drop(['Symbol','SEC Filings','52 Week High','52 Week Low','Name','Sector','Price/Share','Market Cap','EBITDA'],axis=1)
newdata.iplot()
```

```python
df['Number of Shares Outstanding']=df['Market Cap']/df['Price']
```

```python
X=df[['Earnings/Share','EBITDA','Dividend Yield','Number of Shares Outstanding','52 Week High','52 Week Low']]
y=df['Price']
```

```python
df.head()
```

```python
df[['Number of Shares Outstanding','Price']].corr()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
lm.fit(X_train,y_train)
print(lm.intercept_)
```

```python
coeff_df = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficient'])
coeff_df
```

```python
predictions=lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
