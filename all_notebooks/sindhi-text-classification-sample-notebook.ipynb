# Sindhi Text Classification Sample Notebook
#### Here in this notebook we will be creating a simple Decision Tree Classifier for sindhi text classification.
 - The feature taken for the classification is the title of the text
 - Label is the genre of the article i.e. entertainment, sport and technology
 - The Goal is to predict genre by text of title

```python
#importing pandas with alias pd
import pandas as pd
```

```python
#creating a dataframe from the dataset
dataset=pd.read_csv("../input/awamiawaz-sindhi-articles-classification-dataset/Articles Dataset.csv")
```

```python
# having a look at the dataset
dataset.head()
```

```python
# intializing the label column for the classifier
y=dataset.iloc[:,4].values
y
```

```python
# creating corpus of vector from title of article 
corpus=[]
for i in range(0,3366):
    review = dataset['title'][i]
    review=review.split()
    review=' '.join(review)
    corpus.append(review)
```

```python
# creating matrix of features using CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer    
cv=CountVectorizer(max_features=3000)
x=cv.fit_transform(corpus).toarray()
```

```python
# applying label encoder to the y as labels in y are form of text
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y=le.fit_transform(y)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state=0)
```

```python
#importing the deicsion tree classifier
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()
#fiting the classifier on training data
classifier.fit(X_train,y_train)
#performing prediction on testing data
pred=classifier.predict(X_test)
```

```python
#importing the evaluation paramters 
from sklearn.metrics import accuracy_score,precision_score,recall_score
from sklearn.metrics import confusion_matrix
#creating confusion matrix based on the results
cm=confusion_matrix(y_test,pred)
print("confusion matrix")
print(cm)
#printing accuracy of the classifier
print('Accuracy score: {}'.format(accuracy_score(y_test, pred)))
#printing precision of the classifier
print('Precision score: {}'.format(precision_score(y_test, pred,average=None)))
#printing accuracy of the classifier
print('Recall score: {}'.format(recall_score(y_test, pred,average=None)))
```
