# importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing the dataset
df = pd.read_csv('../input/simple-regression/simple_regression.csv')
print(df)
```

```python
print(df.info())
```

```python
# extracting x and y 
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
print(df.isna().any().sum())
```

```python
# splitting x and y into training and test data
from sklearn.model_selection import train_test_split
x_tr,x_te,y_tr,y_te = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# importing and training the linear regression model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_tr,y_tr)
```

```python
# we are going to predict the values
y_pred = lr.predict(x_te)
print(lr.coef_)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse = mean_squared_error(y_te,y_pred)
r2 = r2_score(y_te,y_pred)
print(mse)
print(r2)
```

```python
plt.scatter(x_te,y_te,c='red')
#plt.scatter(x_te,y_pred,c='yellow')
plt.plot(x_te,y_pred,c='blue')
plt.show()
```
