```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('../input/wine-customer-segmentation/Wine.csv',sep = ",", header = 0)
df.head()
Y = df["Customer_Segment"]
```

```python
df.info()
```

```python
df.drop('Customer_Segment', axis = 1, inplace = True)
df.drop('Flavanoids', axis = 1, inplace = True)
```

```python
df.describe()
```

```python
sns.heatmap(df.isnull(), cbar=False)
```

```python
sns.pairplot(df)
```

```python
plt.figure(figsize=(12,10), dpi= 70)
sns.heatmap(df.corr(), cmap='RdYlGn', center=0, annot=True)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_copy = df.copy()
df_copy = pd.DataFrame(scaler.fit_transform(df), 
                                      index=df.index,
                                      columns=df.columns)
df_copy.describe()
```

```python
X1 = df_copy.iloc[: , :].values
inertia = []
for n in range(1 , 11):
    algorithm = (KMeans(n_clusters = n ,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state = 0, algorithm='elkan') )
    algorithm.fit(X1)
    inertia.append(algorithm.inertia_)
plt.figure(1 , figsize = (15 ,6))
plt.plot(np.arange(1 , 11) , inertia , 'o')
plt.plot(np.arange(1 , 11) , inertia , '-' , alpha = 0.5)
plt.xlabel('Number of Clusters') , plt.ylabel('WCSS')
plt.show()
```

```python
algorithm = (KMeans(n_clusters = 3 ,init='k-means++', n_init = 10 ,max_iter=300, 
                        tol=0.0001,  random_state= 0, algorithm='elkan') )
algorithm.fit(X1)
algorithm.cluster_centers_
```

```python
labels = algorithm.labels_ + 1
df['Cluster'] = labels
df.head()
```

```python
sns.pairplot(df[['Proline','OD280','Color_Intensity','Alcohol','Hue','Total_Phenols','Cluster']],palette = 'colorblind',hue='Cluster');
```
