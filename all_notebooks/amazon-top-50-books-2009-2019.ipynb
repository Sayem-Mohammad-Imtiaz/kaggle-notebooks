```python
df = pd.read_csv('/kaggle/input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
df.head()
```

```python
temp_df = df['User Rating'].value_counts().reset_index()
temp_df
```

```python
trace1 = go.Bar(
                x = temp_df['index'],
                y = temp_df['User Rating'],
                marker = dict(color = '#597ff0',
                             line = dict(color='rgb(0,0,0)',width=1.5)))
layout = go.Layout(template = 'plotly_dark', title = 'Ratings Count', xaxis = dict(title = 'Rating'), yaxis = dict(title = 'Count'))
fig = go.Figure(data = [trace1], layout = layout)
fig.show()
```

```python
genre = df.Genre.value_counts().reset_index()
genre
```

```python
trace2 = go.Bar(
                x = genre['index'],
                y = genre['Genre'],
                marker = dict(color = '#597ff0',
                              line = dict(color = 'rgb(0,0,0)', width = 1.5)))
layout = go.Layout(template = 'plotly_dark',xaxis = dict(title = 'Genre'), yaxis = dict(title = 'Count'))
fig = go.Figure( data = [trace2], layout = layout)
fig.show()

def pie_plot(cnt_srs, colors, title):
    labels=cnt_srs.index
    values=cnt_srs.values
    trace = go.Pie(labels=labels, 
                   values=values, 
                   title=title, 
                   hoverinfo='percent+value', 
                   textinfo='percent',
                   textposition='inside',
                   hole=0.7,
                   showlegend=True,
                   marker=dict(colors=colors,
                               line=dict(color='#000000',
                                         width=2),
                              )
                  )
    return trace
plotly.offline.iplot([pie_plot(df.Genre.value_counts(), ['19DF3', 'B6E880'], 'Genre')])
```

```python
df1 = df[df['Genre'] == 'Fiction']
df2 = df[df['Genre'] == 'Non Fiction']

temp_df1 = df1['User Rating'].value_counts().reset_index()
temp_df2 = df2['User Rating'].value_counts().reset_index()

trace1 = go.Bar(
                x = temp_df1['index'],
                y = temp_df1['User Rating'],
                name = 'Fiction',
                marker = dict(color = '#597ff0',
                              line = dict(color = 'rgb(0,0,0)', width = 1.5)))

trace2 = go.Bar(
                x = temp_df2['index'],
                y = temp_df2['User Rating'],
                name = 'Non Fiction',
                marker = dict(color = '#cf59f0',
                              line = dict(color = 'rgb(0,0,0)', width = 1.5)))
layout = go.Layout( template = 'plotly_dark', title = 'Rating By Genre', xaxis = dict(title = 'Rating'), yaxis = dict(title = 'Count'))
fig = go.Figure(data = [trace1,trace2], layout=layout)
fig.show()
```

```python
temp_df1 = df.groupby('Author').count().reset_index().sort_values('Name',ascending=False).head(10)

trace1 = go.Bar(
                x = temp_df1['Author'],
                y = temp_df1['Name'],
                marker = dict(color = '#597ff0',
                              line = dict(color = 'rgb(0,0,0)', width=1.5)))
layout = go.Layout(template = 'plotly_dark',title = 'Top 10 Authors With Most Bestsellers', xaxis = dict(title='Author'), yaxis = dict(title='Count'))
fig = go.Figure(data = [trace1], layout=layout)
fig.show()
```

```python
free = df[df['Price'] == 0]
temp_df1 = free.groupby('Author').count().reset_index().sort_values('Name', ascending=False).head(10)

trace1 = go.Bar(
                x = temp_df1['Author'],
                y = temp_df1['Name'],
                marker = dict(color = '#597ff0',
                              line = dict(color = 'rgb(0,0,0)',width=1.5)))

layout = go.Layout(template = 'plotly_dark', title='Authors of Free Bestsellers', xaxis = dict(title='Author'),yaxis=dict(title='Count'))
fig = go.Figure(data=[trace1],layout=layout)
fig.show()
```

```python
df_2017f = df[(df['Year'] == 2015) & (df['Genre'] == 'Fiction')]
temp_df1 = df_2017f.reset_index().sort_values('Reviews', ascending=False).head(10)
trace1 = go.Bar(
                x = temp_df1['Name'],
                y = temp_df1['Reviews'],
                marker = dict(color = '#597ff0',
                              line = dict(color ='rgb(0,0,0)',width=1.5)))

layout = go.Layout(template='plotly_dark', title = 'Top 10 Fiction Books of 2015 per Reviews', xaxis = dict(title='Name'), yaxis=dict(title='Reviews'))
fig = go.Figure(layout=layout, data=[trace1])
fig.show()
```

```python
temp_df1 = free.groupby('Year').mean().reset_index()



trace1 = go.Scatter(
                    x=temp_df1['Year'], 
                    y=temp_df1['User Rating'], 
                    name="Fiction", 
                    marker=dict(color = '#cf59f0',
                             line=dict(color='rgb(0,0,0)',width=1.5)))


layout = go.Layout(hovermode= 'closest', title = 'Average Rating Of Free Bestsellers Over The Years' , xaxis = dict(title = 'Year'), yaxis = dict(title = 'Average Rating'),template= "plotly_dark")
fig = go.Figure(data = [trace1], layout=layout)
fig.show()
```

```python
top = df[(df['User Rating'] == 4.9) & (df['Reviews'] > 10000)]
temp_df1 = top.groupby('Year').mean().reset_index()

trace1 = go.Scatter(
                    x=temp_df1['Year'], 
                    y=temp_df1['Reviews'], 
                    name="Fiction", 
                    marker=dict(color = '#cf59f0',
                             line=dict(color='rgb(0,0,0)',width=1.5)))


layout = go.Layout(hovermode= 'closest', title = 'Average Reviews of High Rate Bestsellers Over The Years' , xaxis = dict(title = 'Year'), yaxis = dict(title = 'Average Rating'),template= "plotly_dark")
fig = go.Figure(data = [trace1], layout=layout)
fig.show()
```
