##2007's data ****

```python
data_2007 = data[data.Year==2007]

data_2007.head(2)
```

```python
import matplotlib.pylab as plt

plt.scatter(data_2007['Engine Displacement'], data_2007['Combined MPG (FT1)'])


plt.xlabel('Engine Displacement')
plt.ylabel('Fuel Efficiency')
plt.title('The relationship between engine displacement and fuel efficiency of all 2007 model year vehicles ')
```

```python
#The training data

from sklearn.linear_model import LinearRegression

# Create training predictors data
train_X = data_2007['Engine Displacement'].values.reshape(-1, 1)
train_Y = data_2007['Combined MPG (FT1)']
```

```python


reg = LinearRegression().fit(train_X, train_Y)

print("the score of linear regression is : ")
reg.score(train_X, train_Y)
```

```python
data_2008 = data[data.Year==2008]

```

```python
# Create test predictors data

test_X = data_2008['Engine Displacement'][1:200]
test_X = test_X.values.reshape(-1, 1)
```

```python
FE_2008 = reg.predict(test_X)
#The predicted data
y_predicted = FE_2008
```

```python
y_actual = data_2008['Combined MPG (FT1)'][1:200]

plt.scatter(y_predicted, y_actual)

plt.xlabel('Predicted')
plt.ylabel('Observed')
```

```python
from sklearn.metrics import mean_squared_error

from math import sqrt

rms = sqrt(mean_squared_error(y_actual, y_predicted))
print('The Root Mean Squared Error (RMSE) is :')
print(str(rms) + ' which looks like a good result as a first modelling idk')
```

#### Using MultiVariate Adaptive Regression 

```python
from pyearth import Earth

#mars solution
model = Earth()

model = Earth(max_degree=2, penalty=1.0, minspan_alpha = 0.01, endspan_alpha = 0.01, endspan=5) #2nd degree formula is necessary to see interactions, penalty and alpha values for making model simple

model.fit(train_X, train_Y)
model.score(train_X, train_Y)
```

```python
FE_2008_bis = model.predict(test_X)
#The predicted data
y_predicted_mars = FE_2008_bis

plt.scatter(y_predicted_mars, y_actual)

plt.xlabel('Predicted')
plt.ylabel('Observed')


rms = sqrt(mean_squared_error(y_actual, y_predicted_mars))
rms
```
