```python
df = pd.read_csv('/kaggle/input/reddit-rwallstreetbets/r_wallstreetbets_posts.csv')

print(df.columns)
df
```

```python
df = df[['title', 'score', 'author', 'author_flair_text', 'removed_by', 'created_utc']].drop_duplicates()
df = df.drop_duplicates(['title', 'author'])
df
```
