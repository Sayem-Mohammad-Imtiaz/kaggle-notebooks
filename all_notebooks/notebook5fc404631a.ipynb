```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split  
#import matplotlib.pyplot as plt  
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv("../input/flight-take-off-data-jfk-airport/M1_final.csv")
df.head()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
df['OP_UNIQUE_CARRIER']= label_encoder.fit_transform(df['OP_UNIQUE_CARRIER'])#.astype(str)
df['TAIL_NUM']= label_encoder.fit_transform(df['TAIL_NUM'])#.astype(str)
df['DEST']= label_encoder.fit_transform(df['DEST'])#.astype(str)
df['Wind']= label_encoder.fit_transform(df['Wind'].astype(str))
df['Condition']= label_encoder.fit_transform(df['Condition'])#.astype(str)
```

```python
x = df.drop(columns = 'TAXI_OUT',axis=1)
y = df['TAXI_OUT']
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=1)
```

```python
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
#Linear regression
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(x_train, y_train)
y_pred1=model1.predict(x_test)
rmse1=np.sqrt(mean_squared_error(y_test,y_pred1))
print(rmse1)
```

```python
#Ridge regression
from sklearn.linear_model import Ridge
model2 = Ridge(alpha=0.05, normalize=True)
model2.fit(x_train, y_train) 
y_pred2= model2.predict(x_test)
rmse2=np.sqrt(mean_squared_error(y_test,y_pred2))
print(rmse2)
```

```python
#Lasso regression
from sklearn.linear_model import Lasso
model3 = Lasso(alpha=0.03, normalize=True)
model3.fit(x_train, y_train) 
y_pred3= model3.predict(x_test)
rmse3=np.sqrt(mean_squared_error(y_test,y_pred3))
print(rmse3)
```

```python
#KNN model
from sklearn.neighbors import KNeighborsRegressor
model4=KNeighborsRegressor(n_neighbors=100)
model4.fit(x_train,y_train)
y_pred4=model4.predict(x_test)
rmse4=np.sqrt(mean_squared_error(y_test,y_pred4))
print(rmse4)
```

```python
#SVR model
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train_ = sc.fit_transform(x_train)
x_test_ = sc.fit_transform(x_test)
model5=SVR()
model5.fit(x_train_,y_train)
y_pred5=model5.predict(x_test_)
rmse5=np.sqrt(mean_squared_error(y_test,y_pred5))
print(rmse5)
```

```python
#Naive Bayes Regression
from sklearn.linear_model import BayesianRidge
model6=BayesianRidge()
model6.fit(x_train,y_train)
y_pred6=model6.predict(x_test)
rmse6=np.sqrt(mean_squared_error(y_test,y_pred6))
print(rmse6)
```

```python
#Random forest regression
from sklearn.ensemble import RandomForestRegressor
model7=RandomForestRegressor()
model7.fit(x_train,y_train)
y_pred7=model7.predict(x_test)
rmse7=np.sqrt(mean_squared_error(y_test,y_pred7))
print(rmse7)
```

```python
#Light GBM regression
from lightgbm import LGBMRegressor
model8=LGBMRegressor()
x_train__=x_train.drop("Dew Point",axis=1)
x_test__=x_test.drop("Dew Point",axis=1)
model8.fit(x_train__,y_train)
y_pred8 = model8.predict(x_test__)
rmse8=np.sqrt(mean_squared_error(y_test,y_pred8))
print(rmse8)
```
