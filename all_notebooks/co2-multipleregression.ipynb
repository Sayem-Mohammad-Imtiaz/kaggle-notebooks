```python
df=pd.read_csv('/kaggle/input/co2-emission-by-vehicles/CO2 Emissions_Canada.csv')
df.head()
```

```python
df.info()
df.shape
df.columns
```

```python
df.rename(columns={'CO2 Emissions(g/km)' : 'CO2_emission'}, inplace=True)
df.rename(columns={'Engine Size(L)' : 'Enginesize'}, inplace=True)
df.rename(columns={'Fuel Consumption City (L/100 km)' : 'fuelconsumptioncity'}, inplace=True)
df.rename(columns={'Fuel Consumption Hwy (L/100 km)' : 'fuelconsumptionhwy'}, inplace=True)
df.rename(columns={'Fuel Consumption Comb (L/100 km)' : 'fuelconsumptioncomb'}, inplace=True)
df.rename(columns={'Fuel Consumption Comb (mpg)' : 'fuelconsumptioncombmpg'}, inplace=True)
```

```python
df.isna().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.boxplot(x=df.CO2_emission)
plt.title('CO2 Emissions boxplot', fontsize = 20)
plt.show()
```

```python
df.corr().CO2_emission.sort_values()[::-1]
```

```python
sns.heatmap(df.corr(),annot=True, fmt='.0%')
```

```python
from sklearn.model_selection import train_test_split
X=df[["CO2_emission"]]
Y=df[["Enginesize","fuelconsumptioncity","fuelconsumptionhwy","fuelconsumptioncomb","fuelconsumptioncombmpg"]]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(X_train,Y_train)
print(lm.coef_)
print(lm.intercept_)
```

```python
Y_train_pred=lm.predict(X_train)
Y_train_pred
```

```python
Y_test_pred=lm.predict(X_test)
Y_test_pred
```

```python
res=Y_train-Y_train_pred
sns.distplot(res)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(Y_train,Y_train_pred)
print(mse)
```

```python
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(Y_train,Y_train_pred)
print(mae)
```

```python

from sklearn.metrics import mean_squared_error
mse=mean_squared_error(Y_train,Y_train_pred)
rmse=np.sqrt(mse)
print(rmse)
```

```python
from sklearn.metrics import r2_score
r2=r2_score(Y_train,Y_train_pred)
print(r2)
```
