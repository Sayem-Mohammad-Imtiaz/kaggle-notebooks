# Prediction model using TensorFlow v2
**Didn't add fine tuning to improve the accuracy of the model, below code setup is just the demonstration on how to use TensorFlow for data preprocssing and neural network creation
You can fine tune neural netork by optimizing hyper params, hidden layers or try models like Decision tree, KNN or gradient boosing models or ensemble model like Random Forest if your main focus is on accuracy**

```python
y = dataset.pop("y")
y = y.map({"yes": 1, "no": 0})
```

```python
dataset_info = {}
for name, column in dataset.items():
    if column.dtype == object:
        datatype = tf.string
    else:
        datatype = tf.float32
    dataset_info[name] = tf.keras.Input(shape=(1,), dtype=datatype, name=name)

numeric_inputs = {name: value for name, value in dataset_info.items() if value.dtype == tf.float32}
print(numeric_inputs.values())
```

**Normalise the numeric inputs to avoid initial incorrect feature weights**

```python
normalise = preprocessing.Normalization()
normalise.adapt(numpy.array(dataset[numeric_inputs.keys()]))
concat_input = tf.keras.layers.Concatenate()(numeric_inputs.values())
numeric_input_preprocessed = normalise(concat_input)

preprocessed_data = [numeric_input_preprocessed]
print(preprocessed_data)
```

**Encode the categorical fields, object type featues to numeric type as Neural network doesnt process String based input data**

```python
for name, value in dataset_info.items():
    if value.dtype == tf.float32:
        continue

    lookup = preprocessing.StringLookup(vocabulary=dataset[name].unique())
    encoding = preprocessing.CategoryEncoding(max_tokens=lookup.vocab_size())

    lookup_val = lookup(value)
    encoded = encoding(lookup_val)

    preprocessed_data.append(encoded)

print(preprocessed_data)
preprocessed_data_concat = tf.keras.layers.Concatenate()(preprocessed_data)
```

```python
learning_rate = 1e-4
body = tf.keras.Sequential([
    tf.keras.layers.Dense(128, activation=tf.keras.activations.relu, kernel_regularizer='l1'),
    tf.keras.layers.Dense(128, activation=tf.keras.activations.tanh), 
    tf.keras.layers.Dense(128, activation=tf.keras.activations.relu, kernel_regularizer='l1'),
    tf.keras.layers.Dense(1)
])

preprocessed_input = tf.keras.Model(dataset_info, preprocessed_data_concat)
result = body(preprocessed_input(dataset_info))
bank_model = tf.keras.Model(dataset_info, result)
bank_model.compile(loss=tf.keras.losses.BinaryCrossentropy(from_logits=True), optimizer=tf.keras.optimizers.Adam(lr=learning_rate), metrics="accuracy")
```
