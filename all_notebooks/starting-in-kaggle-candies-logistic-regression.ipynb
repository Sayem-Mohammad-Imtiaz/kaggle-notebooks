* sugarpercent: The percentile of sugar it falls under within the data set.

# - Regression Classification

Logistic Regression using a sklearn library.

The creation of the matrix of features (X) already done before.
We still need to drop chocolate off  to a separate vector y, which is the variable that we want to predict knowing the others.


```python
y = X['chocolate']
X.drop(['chocolate'], axis=1, inplace=True)
X
```

```python
# split in train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=1)
```

```python
#apply model
model = LogisticRegression()
model.fit(X_train, y_train)
print(f"Model Accuracy: {model.score(X_test, y_test)}")
```
