## 1. Introduction

In response to the current pandemic, the scientific community requested help to process tens of thousands of research articles related to COVID-19. One subset of research questions relates to the lifespan of the virus. We developed a tool to help researchers identify, among the thousands of research articles, what is known about transmission, incubation, and environmental stability of COVID-19. 

Understanding the duration before an exposed individual begins to show symptoms and the duration for which an individual will remain contagious even after recovery helps formulate quarantine and self-isolating policies. For example, how long should a person in contact with a patient or with travel history to affected regions be quarantined? How long should a recovering patient continue to self-isolate to lower the risk of ongoing transmission? Meanwhile, knowledge regarding the virus's viability in various surfaces and environments is crucial for advising health care facilities, schools, workplaces, and other community organizations on disinfection procedures and the use of personal protective equipment. 

Together with the evaluation of existing preventive strategies, we can assess whether current interventions and policies are supported by evidence and are adequate in curbing the pandemic.

## 2. Software

![](https://drive.google.com/uc?id=1RrYF_32jb4mSZ5BM9iPmAanc1VYPrOoO)

### Approach
To maximize the effectiveness of a tool for experts, we took the approach of building a powerful, intuitive tool designed to tackle a very specific problem, very well. 

First we narrowed our focus to topics with (in our view) the most immediate policy implications:
* Range of incubation periods for the disease in humans (and how this varies across age and health status) and how long individuals are contagious, even after recovery.
* Prevalence of asymptomatic shedding and transmission (e.g., particularly children).
* Effectiveness of movement control strategies to prevent secondary transmission in health care and community settings
* Persistence of virus on surfaces of different materials (e,g., copper, steel, plastic).
* Effectiveness of personal protective equipment and its usefulness to reduce risk of transmission in health care and community settings

Second, we chose a search engine as our tool of interest. While we could not outpace the work of thousands of engineers over decades, at Google, the advantage of our highly narrow topic, and a tool which experts are already highly familiar with, seemed hard to ignore. 

A keyword-based search engine is simple and effective, especially when the research question is specific or involves rarely used domain terminologies. The downside, however, is that the search output can be too narrow when too many conditions are set or too broad when users do not have a clearly defined question. To address the disadvantages, we leveraged newer, more sophisticated context-specific models (SciBERT), with fine-tuning around the essential keywords for our topic, to present a state-of-the-art search experience. 




** Part 1: Process based on topic and Word similarity (BM25) **

The construction of the search engine begins with creating a schema to index all of the thousands of documents. Once that's done, the engine looks for input keywords in the specified fields such as title or abstract and return all articles containing matched keyword ranked by BM25 scores. Our search engine is built with Whoosh, a pure Python search engine library.

** Part 2: Process based on Semantic Relationships ** 

To improve search experience and scope, an LDA model learns correlations between articles based on amount of similar words used across all of the articles. This provides a rough recommendation of similar articles. Instead of using traditional similarity models that do not account for context, we use context-sensitive ones (SciBERT, a BERT model pre-trained on scientific literature) to enable semantic searching. These occur in two places: First, to extract sentences most similar to abstracts, and second to replace context-insensitive word representations traditionally used to train LDA.

** Part 3: Semantically meaningful Search Queries **

Instead of typing out a string of words, that have approximate meanings (like a google search), our queries begin with precisely defined categories, which the model is explicitly tuned to understand. Each question takes on several semantic components relevant to TIES. Consider the following example question:

* “What is the range of incubation periods in humans?”

Instead of a google search, we provide a strict, semantically rich query such as the following:

* ("covid-19" OR "2019-ncov" OR ncov OR "sars-cov-2" OR ...) 

** NOT **

* (“animal” OR “dog” OR “canine” OR “feline” OR ...) 

** AND **

* ("incubation period") 

** OR **

*(“onset” AND “exposure”)




What the software sees is equivalent to the following:


* Target disease: COVID-19 and its synonyms such SARS-Cov-2, novel coronavirus, coronavirus disease 2019 etc
* Population of interest: children, elderly, pregnant women, healthcare workers and so on
* Population not of interest: since the question is about humans, it makes sense to exclude animal related keywords
* Setting: healthcare facilities or community
* Outcome: the specific variable we are trying to evaluate, e.g., incubation periods and its related words such as exposure and symptom onset


** Part 4: Scanning for Relevant Articles **

Now that we have our search queries, initial results are returned from the keyword search engine. For each result, the topic model (Part 2) finds the 20 closest-matched articles. Finally, articles that consistently appear across these lists are filtered to the top. The number of times an article appears determines its overall relevance to the initial search output.


### Proof-of-Concept Data Extraction and Synthesis Tool 

In addition to the search results, we explored the feasibility of extracting study attributes and outcomes using a combination of Natural Language Processing techniques including entity relation extraction and rule-based matching. In this demonstration, we extracted mean or median age from the Methods and Results sections of the papers, as well as incubation periods and prevalence of asymptomatic infection in the Results section. (Due to difficulty in merging metadata with full text files, we were only able to match 53 unique articles with additional attributes.) Full code to feature extractions is documented here.
