```python
import warnings as warnings
warnings.filterwarnings("ignore")

from datetime import datetime, timedelta

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import ScalarFormatter
```

```python
df = data.copy()
df_pricepersales = df[['Name','Price/Sales']]
df_pricepersales = df_pricepersales.sort_values(by='Price/Sales', ascending=True)
```

```python
df_marketcap = df[["Name","Market Cap"]]
df_marketcap = df_marketcap.sort_values(by = "Market Cap",ascending=True).reset_index()
df_marketcap = df_marketcap.tail(10)

plt.figure(figsize=(15,5))
plt.barh(y=df_marketcap['Name'], width=df_marketcap['Market Cap'],edgecolor = "black")
plt.title('Top 10 Companies in S&P 500 Index (By Market Capitalization) ',fontsize = 15)
plt.xlabel(" Market Capitalization (in USD)")
plt.ticklabel_format(useOffset=False, style='plain', axis='x')
plt.show()
```
