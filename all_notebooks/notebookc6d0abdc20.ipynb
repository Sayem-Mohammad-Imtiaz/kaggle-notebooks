# **Champions**
At the first step, we want to know which teams and how many were champions in last seasons We Draw a simple bar chart( Use Seabron )

```python
Champions_team=df_all_tables[df_all_tables['Place']==1]
Champions_team=Champions_team.groupby(by=['Team']).sum()
sns.set_theme(style="whitegrid")
ax = sns.barplot(x=Champions_team.index, y=Champions_team['Place'], data=Champions_team)
```

```python
Goals= df_all_tables.groupby(by=['Team']).sum().sort_values(by=['GA'],ascending=False).drop(['Year','Place'],axis=1)
print("#The Worst defensive line")
print(Goals.iloc[0].name)
Goals= df_all_tables.groupby(by=['Team']).sum().sort_values(by=['GF'],ascending=False).drop(['Year','Place'],axis=1)
print("#The Best Offensive line")
print(Goals.iloc[0].name)
Goals= df_all_tables.groupby(by=['Team']).sum().sort_values(by=['P'],ascending=False).drop(['Year','Place'],axis=1)
print("#The Best  Team")
print(Goals.iloc[0].name)
```
