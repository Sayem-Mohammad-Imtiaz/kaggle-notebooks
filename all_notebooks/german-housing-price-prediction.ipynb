```python
dataset = pd.read_csv('/kaggle/input/german-house-prices/germany_housing_data_14.07.2020.csv')
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
col_to_fill_zero = ['Usable_area', 'Free_of_Relation', 'Bedrooms', 'Bathrooms', 'Floors', 'Furnishing_quality', 'Year_renovated', 'Energy_source', 'Energy_certificate_type', 'Energy_consumption', 'Energy_efficiency_class']

dataset[col_to_fill_zero] = dataset[col_to_fill_zero].fillna(0)
```

```python
dataset.isna().sum()
```

```python
dataset['Year_built'] = dataset['Year_built'].fillna(dataset.Year_built.mean())
dataset['Garages'] = dataset['Garages'].fillna(dataset.Garages.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
```

```python
dataset.isna().sum()
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
dataset.head()
```

```python
X = dataset.drop('Price', axis = 'columns')
y = dataset.Price
```

```python
len(X), len(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
len(X_train), len(X_test)
```

## by using Linear Regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

## by using L1 Regularization

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso()
lasso_reg.fit(X_train, y_train)
y_predict = lasso_reg.predict(X_test)
```

```python
lasso_reg.score(X_train, y_train)
```

```python
lasso_reg.score(X_test, y_test)
```

## by using L2 Regularization

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge()
ridge_reg.fit(X_train, y_train)
y_predict = ridge_reg.predict(X_test)
```

```python
ridge_reg.score(X_train, y_train)
```

```python
ridge_reg.score(X_test, y_test)
```
