Task 4:

For rating feature, The values 'UR' and 'NR' have the same meaning: Unrated and No rated. This value should be unified into only 1 value

ex: replace 'UR' and 'NR' with no rating.

```python
cleanDB=cleanDB.replace(to_replace="NR", value="no rating", regex=True)
cleanDB=cleanDB.replace(to_replace="UR", value="no rating", regex=True)
cleanDB.tail(50)
```

Task 5:

-The duration for TV Shows are in Seasons, but it is in minutes for Movies. You should unify the duration to all be in minutes

```python
cleanDB['duration'].unique()
cleanDB['duration']=cleanDB['duration'].replace('1 Season', value = "480 min")
cleanDB['duration']=cleanDB['duration'].replace('2 Seasons', value = "960 min")
cleanDB['duration']=cleanDB['duration'].replace('3 Seasons', value = "1440 min")
cleanDB['duration']=cleanDB['duration'].replace('4 Seasons', value = "1920 min")
cleanDB['duration']=cleanDB['duration'].replace('5 Seasons', value = "2400 min")
cleanDB['duration']=cleanDB['duration'].replace('6 Seasons', value = "2880 min")
cleanDB['duration']=cleanDB['duration'].replace('7 Seasons', value = "3360 min")
cleanDB['duration']=cleanDB['duration'].replace('8 Seasons', value = "3840 min")
cleanDB['duration']=cleanDB['duration'].replace('9 Seasons', value = "4320 min")
cleanDB['duration']=cleanDB['duration'].replace('10 Seasons', value = "4800 min")
cleanDB['duration']=cleanDB['duration'].replace('11 Seasons', value = "5280 min")
cleanDB['duration']=cleanDB['duration'].replace('12 Seasons', value = "5760 min")
cleanDB['duration']=cleanDB['duration'].replace('13 Seasons', value = "6240 min")
cleanDB['duration']=cleanDB['duration'].replace('14 Seasons', value = "6720 min")
cleanDB['duration']=cleanDB['duration'].replace('15 Seasons', value = "7200 min")
cleanDB['duration']=cleanDB['duration'].replace('16 Seasons', value = "7680 min")
#-------------------------------------------------------------------------------
cleanDB['duration'] = cleanDB['duration'].str.extract(r'(\d+)', expand=False)
cleanDB['duration']=cleanDB['duration'].astype(str).astype(int)
cleanDB['duration']
```
