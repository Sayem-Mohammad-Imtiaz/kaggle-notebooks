```python
import numpy as np
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
# Import LabelEncoder
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

# Importing and Inspecting the Data <a id="1.2"></a>

```python
# Imporitng the File
df=pd.read_csv('/kaggle/input/uciseeds/seeds.csv')
# Checking the shape of the data frame
print(df.shape)
# Printig top few rows
print(df.head(5))
```

# Fitting the model <a id="2"></a>
Here we describe setting the input, out variable and fitting of the model

# Setting up the input and the output variable <a id="2.1"></a>

```python
# Collecting the features and target individually
X=df.iloc[:,0:7]
y=df.iloc[:,7]
```

# Fitting The Naive Bayes Model <a id="2.2"></a>

```python
#Create a Gaussian Classifier
gnb= GaussianNB()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
```

# Accuracy <a id="2.3"></a>

```python
#Accuracy
print('Accuracy = ', gnb.score(X_test, y_test))

#Confusion Matrix
from sklearn.metrics import confusion_matrix
print('\nConfusion matrix')
print(confusion_matrix(y_test, y_pred))

#Classification Report
from sklearn.metrics import classification_report
print('\nClassification Report')
print(classification_report(y_test, y_pred))  
```

```python
# Looking at the model parameters
print(gnb.get_params())
# Checking the likelyhood Table
```

```python
#Create a multinominal Classifier
mnb= MultinomialNB()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
mnb.fit(X_train, y_train)
y_pred = mnb.predict(X_test)
```

```python
#Accuracy
print('Accuracy = ', mnb.score(X_test, y_test))

#Confusion Matrix
from sklearn.metrics import confusion_matrix
print('\nConfusion matrix')
print(confusion_matrix(y_test, y_pred))

#Classification Report
from sklearn.metrics import classification_report
print('\nClassification Report')
print(classification_report(y_test, y_pred))  
```
