Analysys

- The number of high scores is very low compared to "mid" scores (between 68-80);

```python
# Rate by genre

top_10_by_genre = games.groupby('genre')['metascore'].mean().sort_values(ascending=False)/100
top_10_by_genre
```

```python
fig = px.bar(top_10_by_genre, x= top_10_by_genre.index, y=top_10_by_genre.values, labels={'y':'Average Rating Score', 'x':'Genre'})
fig.show()
```

Analysys 

- Sports games has the highest average rating score;
- Even the action games having the highest numbers of games by genre, it still in the tenth place when it comes to average rating score;
- The difference between the first to fourth place is minimum.

```python
sports = games[games['genre'].str.contains('sports')]
sports.sort_values(by = 'metascore', ascending=False).head(10)
```

Analysys

- Once more, the highest metacores come from the 2000s;
- Few football games appear in the top 10.

```python
# Game release per year
games['year'] = pd.DatetimeIndex(games['date']).year
gamerelease = games.groupby('year').count()['titles'].reset_index()
gamerelease.columns = ['Year','Count']

gpyline = px.line(gamerelease, x = 'Year', y = 'Count')
gpyscatter = px.scatter(gamerelease, x = 'Year', y = 'Count')
linescatter = go.Figure(data = gpyline.data + gpyscatter.data)
linescatter.show()
```

```python
# Playstation metascores
ps = games[games['platforms'].str.contains('PlayStation')].sort_values(by='metascore', ascending=False)
ps.head(10)
```

```python
# Xbox metascores
xbox = games[games['platforms'].str.contains('Xbox')].sort_values(by='metascore', ascending=False)
xbox.head(10)
```

```python
# PS vs XBOX metascore average
xboxmean = xbox.groupby('platforms').mean()['metascore'].reset_index()
xboxmean.columns = ['platforms', 'metascore']
psmean = ps.groupby('platforms').mean()['metascore'].reset_index()
psmean.columns = ['platforms', 'metascore']
xmean = px.scatter(xboxmean, x = 'platforms', y = 'metascore')
xmeanline = px.line(xboxmean, x = 'platforms', y = 'metascore')
playmean = px.scatter(psmean, x = 'platforms', y = 'metascore')
playmeanline = px.line(psmean, x = 'platforms', y = 'metascore')
comparison = go.Figure(playmean.data + playmeanline.data + xmean.data + xmeanline.data)
comparison.show()
```
