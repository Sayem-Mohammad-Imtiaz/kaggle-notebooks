```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../input/smart-home-dataset-with-weather-information/HomeC.csv')
df
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df['Solar [kW]'].hist()
```

```python
plt.figure(figsize=(20,5))
plt.plot(df['Solar [kW]'], df['summary'], '.')
```

```python
plt.figure(figsize=(15,5))
plt.plot(df['summary'], df['temperature'], '.')
plt.title('Temperature Range for Diffrent Weathers')
plt.xlabel('Summary (Weather)')
plt.xticks(rotation=90)
plt.ylabel('Temperature')
```

```python
df.info()
```

```python
# Mean Temperature of particular weather
df.groupby('summary').mean()[['temperature', 'apparentTemperature']].plot(kind='bar')
plt.title('Mean Temprature')
plt.ylabel('Temprature')
plt.xlabel('Weather')
```

```python
# Median Temperature of particular weather
df.groupby('summary').median()[['temperature', 'apparentTemperature']].plot(kind='bar')
plt.title('Median Temprature')
plt.ylabel('Temprature')
plt.xlabel('Weather')
```

```python
# Mean Usage in particular weather
df.groupby('summary').mean()['use [kW]'].plot(kind='bar')
```

```python
# Medan Usage in particular weather
df.groupby('summary').median()['use [kW]'].plot(kind='bar')
```
