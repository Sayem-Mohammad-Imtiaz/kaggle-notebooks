```python
from PIL import Image, ImageColor
import matplotlib.pyplot as plt

def hextorgb(hexstr):
    return ImageColor.getcolor("#"+hexstr, "RGB")

plt.rcParams["figure.figsize"] = (12,7)

f = open("/kaggle/input/python-discord-pixels-analysis/pydis_pixels.csv", "r")

lines = f.readlines()

newimg = Image.new("RGB", (160, 90), (0, 0, 0))

for i in range(1, len(lines)):
    line = lines[i].split(",")
    
    newimg.putpixel((int(line[0]), int(line[1])), hextorgb(line[2]))

newimg = newimg.resize((159*3, 89*3), Image.NEAREST)
plt.imshow(newimg)
plt.show()
```
