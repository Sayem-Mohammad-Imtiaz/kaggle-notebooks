```python
df=pd.read_csv("../input/heights-and-weights-dataset/SOCR-HeightWeight.csv")
df.head()
df.shape
df.isnull().sum()
df.shape
train=df[0:12499]
test=df[12500:]
x_train=train.drop('Weight(Pounds)',axis=1)
y_train=train['Weight(Pounds)']
x_test=test.drop('Weight(Pounds)',axis=1)
y_test=test['Weight(Pounds)']
from sklearn.linear_model import LinearRegression
lreg=LinearRegression()
lreg.fit(x_train,y_train)
pred=lreg.predict(x_test)
lreg.score(x_test,y_test)
lreg.score(x_train,y_train)
rmse_test=np.sqrt(np.mean(np.power((np.array(y_test)-np.array(pred)),2)))
rmse_train=np.sqrt(np.mean(np.power((np.array(y_train)-np.array(lreg.predict(x_train))),2)))

```
