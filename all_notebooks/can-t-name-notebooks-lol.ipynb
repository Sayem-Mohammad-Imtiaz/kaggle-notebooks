```python
import pandas as pd
import numpy as np
df = pd.read_csv('../input/have-you-ever-eaten-these-foods/responses.csv')
pd.set_option('display.max_columns', None)
df.head(5)
```

```python
lista = df.columns.values
lista = lista.tolist()
lista.remove('Gender')
lista.remove('Race')
lista.remove('Home location')
lista.remove('Age in years')
print(lista)
```

```python
tried_food = []
for costumer in range(len(df['Age in years'])):
    tried = 0
    for food in range(len(lista)):
        if(df[lista[food]][costumer] == True):
            tried += 1
    tried_food.append(tried)
```

```python
df1 = pd.DataFrame()
df1['Age'] = df['Age in years']
df1['Number of tried foods'] = tried_food
```

```python
df1.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style = 'darkgrid')
plt.figure(figsize = (10,8))
plt.ylabel('Tried foods')
plt.title('Correlation between age and the number of different tried foods')
plt.xlabel('Age')
sns.lineplot(data = df1, x = 'Age', y = 'Number of tried foods')
```
