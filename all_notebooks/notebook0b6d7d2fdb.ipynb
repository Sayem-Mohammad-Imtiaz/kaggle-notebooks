# User Rating Analysis

```python
fig= px.histogram(df,x='User Rating',title='User Rating Analysis')
fig.show()
```

# Popularity based on review ratings

```python
popular = [0,0,0,0]
for i in df['User Rating']:
    if i<3.5:
        df.fillna('Not Much popular')
    if i>=4.8:
        popular[0] += 1
    if i>=4.5:
        popular[1] += 1
    if i>=4.1:
        popular[2] += 1
    else:
        popular[3] += 1
```

```python
x = ['Extermely Popular(4.8-5.0)','Very Popular(4.5-4.7)','Popular(4.1-4.4)','Least Popular']
y = popular
fig = go.Figure([go.Bar(x=x,y=y)])
fig.show()
```

# Popularity based on Genres

```python
d1 = df[df['User Rating'] >=4.1]
fig = px.histogram(d1,x='Genre',title='Popularity based on Genres')
fig.show()
```
