import importance library

```python
import os
import pandas as pd
import numpy as np
from keras import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler
from keras.callbacks import Callback
import matplotlib.pyplot as plt
```

Read Data

```python
data = pd.read_csv('../input/water-potability/water_potability.csv')
```

```python
data.head(5)
```

```python
#Chack null value
data.isnull().sum()
```

```python
#Fill null value
data['ph'] = data['ph'].fillna(data['ph'].mean())
data['Sulfate'] = data['Sulfate'].fillna(data['Sulfate'].mean())
data['Trihalomethanes'] = data['Trihalomethanes'].fillna(data['Trihalomethanes'].mean())
```

```python
data.describe()
```

```python
y = data['Potability']
x = data.drop(['Potability'],axis='columns').values
```

```python
x = StandardScaler().fit_transform(x)
```

```python
model = Sequential([
        Dense(25,input_shape=(9,),activation='relu'),
        Dense(2,activation='sigmoid')    
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
class CustomCallback(Callback):
    def on_epoch_end(self, epoch, logs={}):
        if logs.get('accuracy')>=0.74:
          print('accuracy 74%')
          self.model.stop_training = True

h = model.fit(x,to_categorical(y),epochs=1300,callbacks=[CustomCallback()])
```

```python
plt.plot(h.history['accuracy'],label='accuracy')
plt.plot(h.history['loss'],label='loss')
plt.legend()
```

```python
print(f"min accuracy is : {min(h.history['accuracy'])}")
print(f"max accuracy is : {max(h.history['accuracy'])}")
```
