## Data Description

Column | Description
:---|:---
`id` | The unique ID assigned to every complaint.
`Date` | The date on which the complaint is lodged.
`Product` | The name of the product against which the complaint is lodged.
`Sub Product` | The name of the subproduct mentioned in the complaint.
`Issue` | The issue raised in the complaint.
`Associated company` | The name of the company involved in the complaint.
`State` | The code of the state in which the complaint is lodged.
`Zipcode` | The zip code of the city in which complaint is lodged
`Tags` | The tags provided to the complaint.
`Submitted via` | The medium by which the complaint is submitted.
`Past dispute` | The flag indicates if the customer has disputed in the past. Values are Yes and No.
`Customer demand` | Has the customer asked for monetary compensation? Values are ‘Customer demands compensation’ and ‘Customer does not demand compensation’
`Time allotted` | Time allotted to solve the complaint. (in days)
`Compensation` | Is the complaint eligible for monetary compensation? Values are 0 (no), 1 (yes)
