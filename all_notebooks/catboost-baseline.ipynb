```python
df["test"]=0
test["test"]=1
total=pd.concat([df,test],axis=0)
```

```python
for i in total.select_dtypes("object"):
    lb=LabelEncoder()
    total[i]=lb.fit_transform(total[i])
    df[i]=lb.transform(df[i])
```

```python
train=total[total["test"]==0].copy()
```

```python
test=total[total["test"]==1].copy()
```

```python
ada=CatBoostClassifier(random_seed=42)
ada.fit(train.drop(["target","test"],axis=1),train["target"])
pred=ada.predict(test.drop(["test"],axis=1))
```
