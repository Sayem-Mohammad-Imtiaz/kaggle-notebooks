```python
df = pd.read_csv('/kaggle/input/porn-data/tumbzilla_labels.csv')
file_categories = open('categories.csv','w')
file_tags = open('tags.csv','w')
file_categories_tags = open('categories_tags.csv','w')

for index, row in df.iterrows():
    categories = str(row['categories']).split('__')
    categories_pairs = list(itertools.combinations(categories, 2))
    for category_pair in categories_pairs:
        file_categories.write(category_pair[0]+','+category_pair[1]+'\n')
    tags = str(row['tags']).split('__')
    tags_pairs = list(itertools.combinations(tags, 2))
    for tag_pair in tags_pairs:
            file_tags.write(tag_pair[0]+','+tag_pair[1]+'\n')
    for category in categories:
        for tag in tags:
            file_categories_tags.write(category+','+tag+'\n')
file_categories.close()
file_tags.close()
file_categories_tags.close()
```
