```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
```

```python
df = pd.read_csv('../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
```

```python
df.describe()
```

```python
(df.shape[0]) - (df.drop_duplicates().shape[0])
```

```python
df = df.drop_duplicates()
```

```python
df.info()
```

```python
sns.countplot(x = df.quality)
```

```python
for i in range(0,df.shape[0]):
    if df.iloc[i,11] >= 7:
        df.iloc[i,11] = 2
    elif 5 <= df.iloc[i,11] < 7:
        df.iloc[i,11] = 1
    else:
        df.iloc[i,11] = 0
```

```python
df
```

```python
x = df.drop(['quality'],axis =1)
```

```python
y = df.quality
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
acc = {}
acc['KNN'] = (accuracy_score(y_test,knn.predict(x_test)))
(accuracy_score(y_test,knn.predict(x_test)))
```

```python
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
acc['Random Forest'] = (accuracy_score(y_test,rfc.predict(x_test)))
(accuracy_score(y_test,rfc.predict(x_test)))
```

```python
lgb = LGBMClassifier(learning_rate = 0.4)
lgb.fit(x_train,y_train)
acc['LightGBM'] = (accuracy_score(y_test,lgb.predict(x_test)))
(accuracy_score(y_test,lgb.predict(x_test)))
```

```python
xgb = XGBClassifier(learning_rate = 0.4)
xgb.fit(x_train,y_train)
acc['XGBoost'] = (accuracy_score(y_test,xgb.predict(x_test)))
(accuracy_score(y_test,xgb.predict(x_test)))
```

```python
sns.jointplot(x = acc.keys(),y = acc.values())
```
