#### Importing Libraries


```python
import pandas as pd
import os
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt 
```

#### Importing data w/ os, pandas

```python
df =  pd.read_csv('../input/tipping/tips.csv')
```

#### Sample

```python
df.head(3)
```

### Sort Values by tip

```python
df.sort_values(by=['tip'], ascending=False).head()
```

### Sort Values by total_bill

```python
df.sort_values(by=['total_bill'], ascending= False).head()
```

### Further Look to data

```python
df.info()
```

### Any Corr. between bill and tip?

```python
sns.regplot(x="total_bill", y="tip", data=df, robust=True);
```

### Which gender tips more?

```python
df.groupby(['sex']).mean()['tip']
```

### Does smokers tip more?

```python
df.groupby(['smoker']).mean()['tip']
```

### Any difference between days?

```python
sns.catplot(x="day", y="tip", data=df, height=6, kind="bar", palette="muted")
```

### Any difference between times?

```python
sns.catplot(x="time", y="tip", data=df, height=6, kind="bar", palette="muted")
```

### Where are the tips concentrate?

```python
plt.hist(df['tip'])
plt.xlabel('Dollars')
plt.ylabel('Number of tips')
plt.show()
```

### Total Bills?

```python
plt.hist(df['total_bill'])
plt.xlabel('Dollars')
plt.ylabel('Number of bills')
plt.show()
```
