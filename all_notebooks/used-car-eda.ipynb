```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../input/used-cars-dataset-from-cardekhocom/cardekho_cleaned.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
sns.heatmap(df.corr(),linewidths=1.5)
```

```python
plt.figure(figsize=(14,8))
df.groupby('seller_type') \
        .size() \
        .sort_values(ascending = False) \
        .iloc[:5] \
        .plot.pie(explode=[0,0,1],autopct='%1.1f%%',shadow=True)
plt.ioff()
```

```python
import plotly
import plotly.express as px
fig=px.histogram(df,x="transmission_type",color="transmission_type")
fig.show()
```

```python
model_counts=df.model.value_counts().sort_values(ascending=False).iloc[0:30]
plt.figure(figsize=(20,5))
sns.barplot(model_counts.index,model_counts.values,palette="Purples_r")
plt.ylabel('Counts')
plt.title('Which products are most common?')
plt.xticks(rotation=90);
```
