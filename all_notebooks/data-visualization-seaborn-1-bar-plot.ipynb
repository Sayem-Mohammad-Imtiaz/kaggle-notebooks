```python
percentagePeopleBelowPovertyLevel.replace("-" , 0.0 , inplace = True)

percentagePeopleBelowPovertyLevel["poverty_rate"].value_counts()
```

```python
percentagePeopleBelowPovertyLevel.poverty_rate = percentagePeopleBelowPovertyLevel.poverty_rate.astype(float)

percentagePeopleBelowPovertyLevel.info()
```

```python
places = percentagePeopleBelowPovertyLevel["Geographic Area"].unique()

places
```

```python
ratio = []

for place in places:
    x = percentagePeopleBelowPovertyLevel[percentagePeopleBelowPovertyLevel["Geographic Area"] == place ]
    average = sum(x.poverty_rate) / len(x)
    newData = ratio.append(average)
    

newData2 = pd.DataFrame(
    {"places" : places,
    "poverty" : ratio})
    
index1 = (newData2["poverty"].sort_values(ascending = False)).index.values
newIndex = newData2.reindex(index1)


newIndex.head(10)
```

VISUALIZATION

```python
plt.figure(figsize = (25,15))
sns.barplot(x = newIndex["places"],y = newIndex["poverty"] )

plt.xticks(rotation = 0)
plt.xlabel("PLACES")
plt.ylabel("POVERTY")
plt.title("Poverty Rate By Region")
```

```python
data.replace(["(X)"] ,0.0 , inplace = True)
data.replace(["-"], 0.0 , inplace = True)
```

```python
data["share_white"] = data["share_white"].astype(float)
data["share_black"] = data["share_black"].astype(float)
data["share_native_american"] = data["share_native_american"].astype(float)
data["share_asian"] = data["share_asian"].astype(float)
data["share_hispanic"] = data["share_hispanic"].astype(float)



data.info()

```

```python
places = list(data["Geographic area"].unique())

places
```

```python
share_white = []
share_black = []
share_native_american = []
share_asian = []
share_hispanic = []
```

```python
for place in places:
    x = data[data["Geographic area"] == place]
    
    share_white.append(sum(x.share_white)/len(x))
    share_black.append(sum(x.share_black)/len(x))
    share_native_american.append(sum(x.share_native_american)/len(x))
    share_asian.append(sum(x.share_asian)/len(x))
    share_hispanic.append(sum(x.share_hispanic)/len(x))
    
```

```python
f , ax = plt.subplots(figsize = (15 ,25) )


sns.barplot(x=share_white,y=places,color= "orange" , label = "r" , alpha = 1)
sns.barplot(x = share_black , y=places , color = "blue" , label ="b" ,alpha = 1)
sns.barplot(x = share_native_american , y = places , color = "cyan" ,label="g" ,alpha = 1 )

```

```python
f , ax = plt.subplots(figsize = (15,25))

sns.barplot(x = share_asian , y= places , color = "red" , alpha = 1)
sns.barplot(x = share_hispanic , y=places , color = "g" ,alpha =0.5)
```
