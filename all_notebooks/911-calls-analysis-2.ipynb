```python
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
df = pd.read_csv('../input/911calls-historic-data/911.csv')
```

```python
df['title'].unique()
```

```python
#Now we create a reasons column using a lambda function
df['Reason'] = df['title'].apply(lambda title: title.split(':')[0])
```

```python
df.head(10)
```

```python
df['Reason'].value_counts()
#Here we have seen the common reasons for calling 911
```

```python
sb.countplot(x='Reason',data=df)
```

We have time stamps, and convert them to date time format

```python
type(df['timeStamp'].iloc[0])
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
dp = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day of Week'] = df['Day of Week'].map(dp)
```

```python
sb.countplot(x='Day of Week',data=df,hue='Reason')
```

```python
#In the above graph the legend is in center, which results in the distortion of graph
sb.countplot(x='Day of Week',data=df,hue='Reason')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sb.countplot(x='Month',data=df,hue='Reason')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
#Perform a group by for the missing months
month_m = df.groupby('Month').count()
month_m.head()
```

```python
month_m['twp'].plot().grid(True)
```

```python
sb.lmplot(x='Month', y='twp', data = month_m.reset_index())
```

```python
df['Date']=df['timeStamp'].apply(lambda t: t.date())
plt.figure(figsize = (8,6))
df.groupby('Date').count()['twp'].plot()
plt.tight_layout()
plt.grid(True)
```

```python
#We will now perform groupby operations and visualise based on the reasons of call
plt.figure(figsize = (8,6))
df[df['Reason']=='Traffic'].groupby('Date').count()['twp'].plot()
plt.title('Traffic')
plt.tight_layout()
plt.grid(True)
```

```python
plt.figure(figsize = (8,6))
df[df['Reason']=='Fire'].groupby('Date').count()['twp'].plot()
plt.title('Fire')
plt.tight_layout()
plt.grid(True)
```

```python
plt.figure(figsize = (8,6))
df[df['Reason']=='EMS'].groupby('Date').count()['twp'].plot()
plt.title('EMS')
plt.tight_layout()
plt.grid(True)
```

```python
#For heatmaps ourdata should be in a matrix form, so in order to visualise in that manner we need to restructure our data
hour_day = df.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
hour_day.head()
```

```python
plt.figure(figsize=(15,7.5))
sb.heatmap(hour_day)
```

```python
plt.figure(figsize=(15,7.5))
sb.clustermap(hour_day)
```
