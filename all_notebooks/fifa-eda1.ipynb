# contract dtype id object should be converted into datetime
# also Release Clause has dtype object need to convert in numeric
# but before that need to remove unneccerary letters from that column.

```python
from datetime import datetime
df["Contract Valid Until"] = pd.to_datetime(df["Contract Valid Until"])
df.dtypes
df['Release Clause'] = df['Release Clause'].str.replace("â‚¬", '')
df['Release Clause'] = df['Release Clause'].str.replace("M", '')
df['Release Clause'] = df['Release Clause'].str.replace("K", '')
df['Release Clause'] = pd.to_numeric(df["Release Clause"])
```

```python
res = df.sort_values(by = ["Position", "Potential", "Age", "Wage", "Release Clause"], ascending = [True, False, True, True, True])
res.head(10)
```
