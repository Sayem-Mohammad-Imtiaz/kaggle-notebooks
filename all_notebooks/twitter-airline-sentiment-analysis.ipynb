```python
# Word cloud for negative sentiment

negative_df = tweet_df[tweet_df["airline_sentiment"] == "negative"]
```

```python
negative_words = ' '.join(negative_df['text'])
```

```python
# basic clean word
negative_clean_words = " ".join([word for word in negative_words.split()
                                    if 'http' not in word
                                        and not word.startswith('@')
                                        and word != 'RT'])
```

```python
from wordcloud import WordCloud,STOPWORDS
import matplotlib.pyplot as plt
```

```python
wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color='black',
                      width=3000,
                      height=2500
                     ).generate(negative_clean_words)
```

```python
plt.figure(1,figsize=(12, 12))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
# word cloud for positive

positive_df = tweet_df[tweet_df["airline_sentiment"] == "positive"]
positive_words = ' '.join(negative_df['text'])
positive_clean_words = " ".join([word for word in positive_words.split()
                                    if 'http' not in word
                                        and not word.startswith('@')
                                        and word != 'RT'])

positive_wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color='black',
                      width=3000,
                      height=2500
                     ).generate(positive_clean_words)

plt.figure(1,figsize=(12, 12))
plt.imshow(positive_wordcloud)
plt.axis("off")
plt.show()
```

```python
# word cloud for neutral

neutral_df = tweet_df[tweet_df["airline_sentiment"] == "neutral"]
neutral_words = ' '.join(negative_df['text'])
neutral_clean_words = " ".join([word for word in neutral_words.split()
                                    if 'http' not in word
                                        and not word.startswith('@')
                                        and word != 'RT'])

neutral_wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color='black',
                      width=3000,
                      height=2500
                     ).generate(neutral_clean_words)

plt.figure(1,figsize=(12, 12))
plt.imshow(neutral_wordcloud)
plt.axis("off")
plt.show()
```
