# Data visualizations

```python
# Fixing the size of the plots
from pylab import rcParams

rcParams['figure.figsize'] = 8,8
```

```python
sns.pairplot(data=ds, kind="reg");
```

- The distribution plot of the variables are close to a normal distribution except for the salary feature. If we plot the boxplot, we can notice the number of outliers.
- The features don't show any linear pattern from which a conclusion can be infered.

```python
sns.countplot(x="gender", palette="dark", data=ds);
```

- There are more males than females in this dataset.
- This can cause an imbalance.

```python
sns.countplot(x="ssc_b", hue="gender", palette="dark", data=ds);
```

- Females have more population in Central-based secondary school compared to Other secondary schools. 
- The same trend is for male.
- Overall, the male population in secondary schools are more than females, irresepective of the type of school.

```python
sns.catplot(x="ssc_b", y="ssc_p", hue="gender", kind="violin", palette="dark", data=ds);
```

- In Others secondary schools, we can make the following observations:
    - Male tend to achieve average scores from 55 to 75, whereas, females are high scorers from 75 to 85.
    - The number of males who are high scores are less than female.
    
- In Central-based secondary schools, we can make the following observations:
    - Again, women tend to score higher from 72 to 82.
    - Males tend to score less than the females from 60 to 70.

```python
sns.catplot(x="hsc_b", y="hsc_p", kind="box", hue="gender",palette="dark", data=ds);
```

- The boxplot clearly shows presence of outliers.
- High school percentage of male is better in Others compared to females. However, in Central, females score more than the female.
- If we compare the school types, both male and female students perform better in Central.

```python
sns.barplot(x="degree_t", y="degree_p", hue="gender", palette="dark", data=ds);
```

- There is more population of female students in degree-level education compared to male students.
- More students tend to choose Sci&Tech compared to the Comm&Mgmt stream.
- The population of female in individual streams of the degrees is higher.

```python
sns.catplot(y="workex", hue="gender", kind="count", palette="dark" ,data=ds);
```

- Overall, most of the students don't tend to have work experience.
- More population of male have no experience compared to female.

```python
sns.catplot(x="specialisation", y="mba_p", hue="gender", kind="point", markers=["^", "o"], linestyles=["-","--"], palette="dark", data=ds);
```

- Female again wins the race by scoring higher than male students.
- The overall percentage score is higher in Mkt&Fin compared to Mkt&HR for both the group of students.

```python
sns.countplot(x="status", hue="gender", palette="dark", data=ds);
```

- More male students are placed compared to female students.
- The no. of female students placed if half the amount of the no. of male students placed.
- In Not Placed, this proportion of difference between male and female Not Placed students is not as high as in the Placed scanrio.

```python
# Only considering the placed student
placed = ds[ds["status"]=="Placed"][["gender","salary", "workex"]]

sns.catplot(y="salary", x="gender", kind="swarm", palette="dark" ,data=placed);
```

- Considering only the placed students, we can make the following observations:
    - Male students tend to achieve the higher salary package compared to female students.
    - Within the range of 2L to 3L, the population of male students is more than female.
    - Female students tend to achieve a salary of 2L to 2.5L more often.

```python
sns.catplot(x="workex", y="salary", hue="gender", kind="swarm", palette="dark" ,data=placed);
```

- Though male students have less work experience, they achieve the higher salary package, compared to female students.
- Considering only the experienced students, male tend to win the race.
- Female students with experience tend to achieve a salary of 2L to 3L only, and very rarely more than 5L. This is not the same for the male students.
- Considering only inexperienced students, male students acheive a salary from 4L to 5L. Whereas, female students lie between 2L to 3L. This is the same range as it was for female students with experience.
