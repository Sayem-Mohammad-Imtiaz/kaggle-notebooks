```python
data.groupby('Type').count().plot.bar()
```

According to the bar graph above we can say that dataset is balanced.
Now we can analyze the correlation of numeric features with corr function.

```python
data.corr()
```

According to the correlation graph above we can say that some features are dependent each other to a certain extent. Temperature and R, L and A_M, L and Type, R and A_M, R and Type seem somewhat dependent on each other. The most important relationship emerges between Type and A_M. Minus sign means that this relationship is inversely proportional. This relationship is most important because correlation value of this two feature is -0.955276. The value is too close to -1 and that means these two values are very dependent to each other.

```python
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt='.1f', ax=ax)
```

Now we can visualize some properties of dataset.

```python
data.columns
```

```python
# Line Plots

fig, axs = plt.subplots(2,2, figsize=(15,15))

data.Temperature.plot(kind='line', color='g', label='Temperature', linewidth=1, alpha=0.5, linestyle='-', ax=axs[0,0])
data.L.plot(kind='line', color='r', label='L', linewidth=1, alpha=0.5, linestyle='-', ax=axs[0,1])
data.R.plot(kind='line', color='b', label='R', linewidth=1, alpha=0.5, linestyle='-', ax=axs[1,0])
data.A_M.plot(kind='line', color='y', label='A_M', linewidth=1, alpha=0.5, linestyle='-', ax=axs[1,1])

axs[0,0].legend(loc='upper right')
axs[0,0].set_xlabel('Index')
axs[0,0].set_ylabel('Temperature')
axs[0,1].legend(loc='upper right')
axs[0,1].set_xlabel('Index')
axs[0,1].set_ylabel('L')
axs[1,0].legend(loc='upper right')
axs[1,0].set_xlabel('Index')
axs[1,0].set_ylabel('R')
axs[1,1].legend(loc='upper right')
axs[1,1].set_xlabel('Index')
axs[1,1].set_ylabel('A_M')


```

```python
# Scatter plot x=type, y=A_M
data.plot(kind='scatter', x='Type', y='A_M', alpha=0.5, color='red')
plt.xlabel('Type')
plt.ylabel('A_M')
plt.title('Type vs A_M')
```

As we can see from the scatter graph above about relation with A_M and Type, when type number of star increases (from 0 to 5 the size of star increases) A_M decreases.

```python
# Histogram
# bins = number of bar in figure
data.A_M.plot(kind='hist', bins=50, figsize=(12,12), grid=True)
```
