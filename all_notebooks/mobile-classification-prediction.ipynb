```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df_mob = pd.read_csv('/kaggle/input/mobile-price-classification/train.csv')
```

```python
df_mob.head()
```

```python
# There are no null values in the dataset 
print(df_mob.isnull().sum())
```

```python
# As seen below the price can be either 0 1 2 3 (4 types of categories)
df_mob['price_range'].value_counts()
```

```python
plt.figure(figsize=[15,8])
plt.subplot(221)
sns.boxplot(x='price_range',y='ram',data=df_mob)
plt.subplot(222)
sns.boxplot(x='four_g',y='ram', hue='price_range',data=df_mob)
plt.subplot(223)
sns.scatterplot(x='battery_power',y='ram', hue='price_range',data=df_mob)
plt.subplot(224)
sns.boxplot(x='touch_screen',y='ram', hue='price_range',data=df_mob)
```

```python
X = df_mob.drop(['price_range'], axis=1).values
y = df_mob['price_range'].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC 
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

```python
models = [['Logistic Regresssion :',LogisticRegression()],
           ['Decision Tree Classification :', DecisionTreeClassifier()],
           ['RandomForest Classifier :',RandomForestClassifier(n_estimators = 200,criterion = 'entropy')],
           ['KNeighbours Classifier:', KNeighborsClassifier(n_neighbors = 5)],
           ['GradientBoost Classifier:', GradientBoostingClassifier(n_estimators = 999)],
           ['SVM :', SVC()],
           ['NB Classifier :', GaussianNB()]]
```

```python
for name_of_model,model_parms in models:
    curr_model = model_parms
    curr_model.fit(X_train, y_train)
    curr_predict = curr_model.predict(X_test)
    print(50*"=")
    print(20*" ",name_of_model,20*" ")
    print(50*"=")
    print("Accuracy Score  :", accuracy_score(y_test, curr_predict))
    print("Confusion Matrix :\n",confusion_matrix(y_test, curr_predict))
    print("Classification Report :\n", classification_report(y_test,curr_predict))
```

```python
# Running the data through Gradient Boosting Classifier
from sklearn.ensemble import GradientBoostingClassifier
grd = GradientBoostingClassifier(n_estimators = 999).fit(X_train,y_train)
```

```python
y_predict = grd.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.metrics import classification_report
cm = confusion_matrix(y_test,y_predict)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test, y_predict))
```
