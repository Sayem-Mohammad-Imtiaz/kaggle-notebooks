```python
from sklearn.impute import SimpleImputer
# data loading, exploration, and cleaning
development = pd.read_csv("/kaggle/input/gender-development-inequality/gender_development_index.csv")
inequality = pd.read_csv("/kaggle/input/gender-development-inequality/gender_inequality_index.csv")
# remove spaces and make it easier to reference numerical columns
development.rename(columns={'HDI Rank (2015)': 'HDI_Rank', 
                            '2000': 'Year_2000',
                            '2005': 'Year_2005',
                            '2010': 'Year_2010',
                            '2011': 'Year_2011',
                            '2012': 'Year_2012',
                            '2013': 'Year_2013',
                            '2014': 'Year_2014',
                            '2015': 'Year_2015'}, inplace=True)
inequality.rename(columns={'HDI Rank (2015)': 'HDI_Rank',
                            '1995': 'Year_1995',
                            '2000': 'Year_2000',
                            '2005': 'Year_2005',
                            '2010': 'Year_2010',
                            '2011': 'Year_2011',
                            '2012': 'Year_2012',
                            '2013': 'Year_2013',
                            '2014': 'Year_2014',
                            '2015': 'Year_2015'}, inplace=True)
# drop the columns that only contain NaN
development.dropna(axis=1, how='all', inplace=True)
inequality.dropna(axis=1, how='all', inplace=True)
# this column also contains ' a' for some reason but it's not useful
inequality.drop('Unnamed: 3', axis=1, inplace=True)
# this row shouldnt have been included in the data, the formatting's all wrong
inequality.drop(index=160, inplace=True)
# set the index to the hdi rank and sort by it
development.set_index("HDI_Rank", inplace=True)
inequality.set_index("HDI_Rank", inplace=True)
development.sort_index(inplace=True, key=lambda l:list(map(int, l)))
inequality.sort_index(inplace=True, key=lambda l:list(map(int, l)))
# print(development.columns)
# print(inequality.columns)
print(development)
print(inequality)
```

```python
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from matplotlib.ticker import MultipleLocator
sns.set_theme(rc={'axes.facecolor':'(0.15, 0.169, 0.208)', 
                  'figure.facecolor':'(0.126, 0.129, 0.141)', 
                  'axes.labelcolor':'.85', 'text.color': '.85', 
                  'xtick.color':'.85', 'ytick.color':'.85',
                  'figure.figsize': (11.7, 8.27)})

# a visual representation of the locations of every dog, color-coded by neighborhood
fig, ax = plt.subplots(figsize = (11.7, 8.27))
ax.xaxis.set_major_locator(MultipleLocator(10))
ax.set_xlabel('HDI Rank')
ax.set_ylabel('Inequality') 
for col in reversed(inequality.columns):
    if "Year" in col:
        plt.title("Gender inequality by HDI index")
        ax.plot_date(x=inequality.index, y=inequality[col], label=col.split('_')[1])
        ax.legend()
```

```python
fig, ax = plt.subplots(figsize = (11.7, 8.27))
ax.xaxis.set_major_locator(MultipleLocator(10))
ax.set_xlabel('HDI Rank')
ax.set_ylabel('Inequality')  
plt.title("Gender inequality by HDI index (2015)")
sns.regplot(x=list(map(int,inequality.index)), y=list(map(float,inequality['Year_2015'])))
```
