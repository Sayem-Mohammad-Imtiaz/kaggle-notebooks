# Create a global map for COVID Vaccinations (percentage of the population that are fully vaccinated).

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
print(df.columns)
print(df)
fig = px.choropleth(df,locations='country',locationmode='country names',color='people_fully_vaccinated_per_hundred',hover_name='country',animation_frame='date',color_continuous_scale='rainbow',range_color=(0.0,80.0))
fig.update_layout(title="Percentage of the country's total population that are fully vaccinated",title_x=0.45)
fig.show()

plotly.offline.plot(fig,filename='vaccinations.html')
```

## Analysis
### From Jan. 19th to recent days, The percentage of the USA is raising.
### Here is the report about this
### https://covid.cdc.gov/covid-data-tracker/#vaccination-trends


# Create a global map of earthquake occurrences worldwide.

```python
df = pd.read_csv('../input/20080512-wenchuan-earthquake/Wenchuan_earthquake.csv',header=0)
df.index = pd.to_datetime(df['time'])
df['time'] = df.index.strftime('%Y-%m-%d %H:00:00')
df = df.sort_index()
fig = px.scatter_geo(df,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='rainbow',range_color=(1.0,9.0))
fig.show()
```
