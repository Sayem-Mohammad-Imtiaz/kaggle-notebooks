### Cleaning

```python
# Import Data
df = pd.read_csv('../input/stock-exchange-data/indexData.csv')
```

```python
# Look for null values
df.isnull().sum()
```

```python
# Drop rows with null values
df.dropna(subset = ["Open"], inplace=True)
df.isnull().sum()
```

### Currency Conversions

```python
# Current Fx Rates (How many USD per one unit of foreign currency)
usdhkd = .13
usdcny = .16
usdjpy = .01
usdeur = 1.22
usdcad = .83
usdinr = .01
usdkrw = .001
usdchf = 1.11
usdtwd = .04
usdzar = .07
```

```python
# Split each index into separate dataframe, add a column for closing price in terms of USD
hsi = df.loc[df['Index'] == 'HSI']
hsi['CloseUSD'] = hsi['Close'] * usdhkd
nya = df.loc[df['Index'] == 'NYA']
nya['CloseUSD'] = nya['Close'] * 1 #Already in USD
ixic = df.loc[df['Index'] == 'IXIC']
ixic['CloseUSD'] = ixic['Close'] * 1 #Already in USD
shanghai = df.loc[df['Index'] == '000001.SS']
shanghai['CloseUSD'] = shanghai['Close'] * usdcny
shenzhen = df.loc[df['Index'] == '399001.SZ']
shenzhen['CloseUSD'] = shenzhen['Close'] * usdcny
n225 = df.loc[df['Index'] == 'N225']
n225['CloseUSD'] = n225['Close'] * usdjpy
n100 = df.loc[df['Index'] == 'N100']
n100['CloseUSD'] = n100['Close'] * usdeur
gdaxi = df.loc[df['Index'] == 'GDAXI']
gdaxi['CloseUSD'] = gdaxi['Close'] * usdeur
gsptse = df.loc[df['Index'] == 'GSPTSE']
gsptse['CloseUSD'] = gsptse['Close'] * usdcad
nsei = df.loc[df['Index'] == 'NSEI']
nsei['CloseUSD'] = nsei['Close'] * usdinr
ks11 = df.loc[df['Index'] == 'ks11']
ks11['CloseUSD'] = ks11['Close'] * usdkrw
ssmi = df.loc[df['Index'] == 'SSMI']
ssmi['CloseUSD'] = ssmi['Close'] * usdchf
twii = df.loc[df['Index'] == 'TWII']
twii['CloseUSD'] = twii['Close'] * usdtwd
j203 = df.loc[df['Index'] == 'J203.JO']
j203['CloseUSD'] = j203['Close'] * usdzar
```

```python
#Recombine the dataframes
frames = [hsi, nya, ixic, shanghai, n225, n100, shenzhen, gsptse, nsei, gdaxi, ks11, ssmi, twii, j203]
df = pd.concat(frames)
```

```python
df.to_csv('indexProcessed.csv',index=False)
```
