## Task 10

How does the adolescent birth rate in a country affect life expectancy?

```python
adolescent_brate = pd.read_csv('/kaggle/input/who-worldhealth-statistics-2020-complete/adolescentBirthRate.csv')
```

```python
new_column_names = ['location', 'period', 'indicator', 'adolescent_birth_rate']
adolescent_brate.columns = new_column_names
```

```python
period_mask = adolescent_brate['period'] == 2015
adolescent_brate_2015 = adolescent_brate[period_mask]
```

```python
columns_to_drop = ['period', 'indicator']
adolescent_brate_2015.drop(columns = columns_to_drop, inplace = True)
```

```python
adolescent_brate_2015.describe()
```

```python
adolescent_brate_2015.info()
```

```python
merged_df = life_expectancy_2015.merge(adolescent_brate_2015, how = 'inner', on = 'location')  # life_expectancy_2015 and adolescent_brate_2015 can be in either order for inner join
```

```python
x_data = merged_df['adolescent_birth_rate']
y_data = merged_df['life_expectancy_at_birth']
plt.scatter(x_data, y_data)
plt.xlabel('Adolescent birth rate /n(per 1000 women aged 15-19 years)')
plt.ylabel('Average life expectancy')
plt.title('Variation of average life expectancy with \nadolescent birth rate per 1000 women aged 15-19')
plt.show()
```

```python
# Split your data into training and testing data

X_train, X_test, y_train, y_test = split_dataset(merged_df, 'adolescent_birth_rate', 'life_expectancy_at_birth')

# Train your machine learning model

linear_model.fit(X_train, y_train)

# Get your predicted values

predictions = linear_model.predict(X_test)
```

```python
mae = mean_absolute_error(y_test, predictions)
mae  # 3.747
```

```python
print(linear_model.coef_)  # -0.141
print(linear_model.intercept_)  # 80.40
```

What do you notice that is different about this model compared to the previous one? Look at the plot and the MAE.

Our observations:
* MAE is lower i.e. predictions closer to true value
* The gradient of the line is negative i.e. adolescent birth rate negatively impacts life expectancy.
