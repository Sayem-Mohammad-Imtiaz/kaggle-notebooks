```python
import plotly as plt
import plotly.graph_objects as go #Visualization
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px




# Get Data: 
data = pd.read_csv('/kaggle/input/top100polutionmay2021/May2021Top100Polution.csv', encoding="ISO-8859-1")


data.describe()

start,end = 0,100 

fig = go.Figure(data=go.Scatter3d(
    x=data['Exp Pollution Index'][start:end],
    y=data['Pollution Index'][start:end],
    z=data['Country'][start:end],
    text=data['Country'][start:end],
    mode='markers',
    marker=dict(
        sizemode='diameter',
        sizeref=2.5,
        size=data['Exp Pollution Index'][start:end],
        color = data['Exp Pollution Index'][start:end],
        colorscale = 'Viridis',
        colorbar_title = 'Exp Pollution Index<br>',
        line_color='rgb(140, 140, 170)'
    )
))








fig.update_layout(height=600, width=600,
                  title='Polution index by Country May 2021 - The Top 100')






fig.show()


#describe data
data.describe()


data.sort_values(by='Pollution Index', ascending=True).head(10)
```
