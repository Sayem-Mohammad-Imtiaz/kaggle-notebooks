```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../input/mbti_1.csv')
```

<h3>Generating features: Average words per post.</h3>
(Burrowed from Dlao who made [this](https://www.kaggle.com/laowingkin/mbti-study-personality) kernel)

```python
def var_row(row):
    l = []
    for i in row.split('|||'):
        l.append(len(i.split()))
    return np.var(l)

df['words_per_comment'] = df['posts'].apply(lambda x: len(x.split())/50)
df['variance_of_word_counts'] = df['posts'].apply(lambda x: var_row(x))
df['http_per_comment'] = df['posts'].apply(lambda x: x.count('http')/50)
df['qm_per_comment'] = df['posts'].apply(lambda x: x.count('?')/50)
df['INTJMENTIONS_per_comment'] = df['posts'].apply(lambda x: x.count('INTJ')/50)
df['ENFPMENTIONS_per_comment'] = df['posts'].apply(lambda x: x.count('ENFP')/50)
df.head()
```

<h2>How many of each type is present in the dataset?</h2>


```python
a4_dims = (9, 6)
fig, ax = plt.subplots(figsize=a4_dims)
sns.countplot(ax=ax,x=df['type'])
plt.show()
```

Seems like we <u>***need to be careful***</u> when comparing difference across mbti-types in this dataset. 

<h2>Histograms: Words per comment vs. Type</h2>
As plotted below, it seems like the IN-types are fairly right-skewed across all four types. <br>
In fact, it looks like the IN-types has the most identical distributions.

```python
df.hist(column='words_per_comment',by = "type", grid=False, bins=20
        ,xlabelsize=8, ylabelsize=8,figsize = (15,15)
       ) #
plt.show()
```

<h1>Histograms: Questionmarks per comment</h1>
(Please remark x-tickmarks are in the bottom histograms)
Visually, the distributions looks farily similar.

```python
#df_2 = df[~df['type'].isin(['ESFJ','ESFP','ESTJ','ESTP'])]
df.hist(column='qm_per_comment',by = "type", grid=False, bins=20
        ,xlabelsize=10, ylabelsize=8,figsize = (15,15), sharex=True
       ) #
plt.show()
```

Hard to determine the difference visually this way. Lets look at the means instead.

```python
Means = df.groupby(['type'], as_index=False).mean()
Means = pd.DataFrame(Means)
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.barplot(ax=ax, x=df['type'], y=df['qm_per_comment'])
plt.show()
```

Very similarly distributed, but means are different. ENTJ, ESTP and ESFP have the most questionmarks. However, these particular types are also minorities in the dataset (n per type is < 250).

<h2>URL's per post:</h2>

```python
df.hist(column='http_per_comment',by = "type", grid=False, bins=20
        ,xlabelsize=10, ylabelsize=8,figsize = (15,15), sharex=True
       ) #
plt.show()
```

Again hard to determine using histograms. However, IS-types seems to have the most variation in terms of distribution shape. Proceeding with barcharts with means again.

```python
Means = df.groupby(['type'], as_index=False).mean()
Means = pd.DataFrame(Means)
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.barplot(ax=ax, x=df['type'], y=df['http_per_comment'])
plt.show()
```

ISFP, ISTP and INFP does on average **include links in their posts most often**.

<h1>INTJ and ENFP mentions </h1>

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.barplot(ax=ax, x=df['type'], y=df['INTJMENTIONS_per_comment'])
plt.show()
```

Apart from the INTJ's themselves, ENTJ, ENFP and ESFP seems to mention the INTJ-type most often.

<h1>The next-most interesting part: ENFP mentions </h1>

```python
a4_dims = (11.7, 8.27)
fig, ax = plt.subplots(figsize=a4_dims)
sns.barplot(ax=ax, x=df['type'], y=df['ENFPMENTIONS_per_comment'])
plt.show()
```

ESFP, ENTP and ENFJ are most often mentioning the ENFP's apart from themselves.
