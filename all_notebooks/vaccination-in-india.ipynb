```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
vaccine = pd.read_csv('../input/covid19-vaccines-and-cases-dataset-state-wise/Covid vaccines and Cases Dataset Statewise (states  UTs).csv')
```

```python
vaccine.head()
```

```python
vaccine.columns
```

# Plotting using the Seaborn Library

```python
vaccine_5crores = vaccine[vaccine['Total Vaccine Doses']>50000000]
```

```python
vaccine_5crores
```

```python
plt.figure(figsize=(15,9))
sns.barplot(x='TOTAL CONFIRMED',y='Total Vaccine Doses',data=vaccine_5crores,hue='State Name')
```

****From the Above Barplot, we can say that Uttar Pradesh and Manipur has taken more number of vaccinations in India. It was around 8,00,00,000 doses and 6,00,00,000+ doses.****

```python
vaccine_lakhs=vaccine[vaccine['Total Vaccine Doses']<100000]
```

```python
vaccine_lakhs
```

```python
vaccine_1crores= vaccine[vaccine['Total Vaccine Doses']>10000000]
```

```python
plt.figure(figsize=(15,9))
plt.xticks(rotation=90)
ax = sns.barplot(x='State Name',y='Total Vaccine Doses',data=vaccine_1crores)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.barplot(x='State Name',y='TOTAL CONFIRMED',data=vaccine)
```

```python
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.barplot(x='State Name',y='DEATH',data=vaccine)
```
