```python
x = df1.drop(columns=['Acession ID','Host','Divided_seq'])
y = pd.DataFrame(df1['Host'])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
clf = SVC(1)
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
```

```python
print(f"SVM accuracy :{accuracy_score(y_pred, y_test)}")
```
