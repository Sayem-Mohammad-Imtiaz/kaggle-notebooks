# Importando e analisando o conjunto de dados

```python
df_train = pd.read_csv('../input/mini-flight-delay-prediction/flight_delays_train.csv')
df_test = pd.read_csv('../input/mini-flight-delay-prediction/flight_delays_test.csv')

df_train
```

# Pr√©-processamento

```python
def to_number(x):
    y = int(re.sub(r"c-", "", x))
    return y

def y_or_n(x):
    if x == "Y":
        return 0
    else:
        return 1
```

## Dados de treino

```python
# Month, DayofMonth e DayofWeek

df_train.Month = df_train.Month.map(lambda x: to_number(x))
df_train.DayofMonth = df_train.DayofMonth.map(lambda x: to_number(x))
df_train.DayOfWeek = df_train.DayOfWeek.map(lambda x: to_number(x))

# UniqueCarrier

uc_categories = {}
i = 0
for x in df_train.UniqueCarrier:
    if x not in uc_categories:
        uc_categories[x] = i
        i += 1
        
uc_categories_int = []
j = 0
for x in df_train.UniqueCarrier:
    uc_categories_int.append(uc_categories[x])
    
df_train.UniqueCarrier = uc_categories_int

# Origin

origin_categories = {}
i = 0
for x in df_train.Origin:
    if x not in origin_categories:
        origin_categories[x] = i
        i += 1
        
origin_categories_int = []
j = 0
for x in df_train.Origin:
    origin_categories_int.append(origin_categories[x])

df_train.Origin = origin_categories_int

# Dest

dest_categories = {}
i = 0
for x in df_train.Dest:
    if x not in dest_categories:
        dest_categories[x] = i
        i += 1
        
dest_categories_int = []
j = 0
for x in df_train.Dest:
    dest_categories_int.append(dest_categories[x])
    
df_train.Dest = dest_categories_int

# dep_delayed_15min

df_train.dep_delayed_15min = df_train.dep_delayed_15min.map(lambda x: y_or_n(x))
```

## Dados de teste

```python
# Month, DayofMonth e DayofWeek

df_test.Month = df_test.Month.map(lambda x: to_number(x))
df_test.DayofMonth = df_test.DayofMonth.map(lambda x: to_number(x))
df_test.DayOfWeek = df_test.DayOfWeek.map(lambda x: to_number(x))

# UniqueCarrier

uc_categories = {}
i = 0
for x in df_test.UniqueCarrier:
    if x not in uc_categories:
        uc_categories[x] = i
        i += 1
        
uc_categories_int = []
j = 0
for x in df_test.UniqueCarrier:
    uc_categories_int.append(uc_categories[x])
    
df_test.UniqueCarrier = uc_categories_int

# Origin

origin_categories = {}
i = 0
for x in df_test.Origin:
    if x not in origin_categories:
        origin_categories[x] = i
        i += 1
        
origin_categories_int = []
j = 0
for x in df_test.Origin:
    origin_categories_int.append(origin_categories[x])

df_test.Origin = origin_categories_int

# Dest

dest_categories = {}
i = 0
for x in df_test.Dest:
    if x not in dest_categories:
        dest_categories[x] = i
        i += 1
        
dest_categories_int = []
j = 0
for x in df_test.Dest:
    dest_categories_int.append(dest_categories[x])
    
df_test.Dest = dest_categories_int
```

# Testando o classificador escolhido

```python
labels = df_train.dep_delayed_15min
df_train.drop(columns="dep_delayed_15min", inplace=True)
features = df_train

train, test, train_labels, test_labels = train_test_split(features, labels, test_size=0.33, random_state=42)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(500, 20), max_iter=3000)
mlp.fit(train, train_labels)

predicted_labels = mlp.predict(test)
print('F1-score: ', metrics.f1_score(test_labels, predicted_labels))
```
