```python
df['disability'].fillna(value = 0, inplace = True)
```

```python
df['race'].fillna(value = 'Unrecorded', inplace = True)
```

```python
df['rating'].fillna(value = 'Unknown', inplace = True)
```

```python
df.gender.value_counts().plot(kind = 'bar',title = 'Gender of Characters',figsize = (10,7))
```

```python
df.role.value_counts().plot(kind = 'bar', title = 'Role of Characters', figsize = (10,7))
```

```python
df.studio.value_counts().plot(kind = 'bar', title = 'Studio', figsize = (15,10))
#plt.xticks(rotation = 45)
plt.show()
```

```python
plt.subplot(221)
df.representation.value_counts().plot(kind = 'bar', title = 'Representation', figsize = (15,10))
plt.xticks(rotation = 45)
plt.subplot(222)
df.orientation.value_counts().plot(kind = 'bar', title = 'Orientation of Characters')
plt.xticks(rotation = 45)
plt.show()
```

```python
a = df['year'].value_counts()
a = pd.DataFrame(a)
a.reset_index(level = 0, inplace = True)
print(a)
```

```python
a.columns = ['year','year_count']
```

```python
a.sort_values(['year'], inplace = True)
```

```python
a
```

```python
plt.plot(a.year, a.year_count)
plt.xticks(rotation = 45, ticks = list(a.year),labels = list(a.year))
plt.show()
```
