```python
import pandas as pd
import numpy as np
from datetime import datetime
import arrow

def dateparse(dt):
    dt=dt.rstrip()
    return datetime.strptime(dt, '%Y %m')
```

```python
#抓出資料，並將兩個資料合併
miaoli_2019 = pd.read_csv("../input/miaoli-wifi-20192020/2019-wifi.csv")
miaoli_2020 = pd.read_csv("../input/miaoli-wifi-20192020/2020-wifi.csv")

miaoli_wifi=pd.concat([miaoli_2019,miaoli_2020])
miaoli_wifi
```

```python
miaoli_wifi.info
```

```python
miaoli_wifi.columns
```

```python
#因為欄位是亂碼，修改欄位名稱
miaoli_wifi.columns=['time','使用人次','使用人數','total_mins','total_GB','download_GB','upload_GB']
miaoli_wifi.info()
```

```python
#將time欄位轉為datetime
miaoli_wifi['time']=pd.to_datetime(miaoli_wifi['time'])
miaoli_wifi.info()
```

```python
#以季做時間區隔，呈現每一季的資料
start = arrow.get('2019-01')
end = start.shift(months=23)
print(f"{start} {end}")
sum_data = []
for r in arrow.Arrow.interval('month', start, end, 3):
    start_date = str(r[0].date())
    end_date  = str(r[1].date()) 
    print (f"{start_date} ,{end_date} ")
    mdf = miaoli_wifi[miaoli_wifi['time'].between(start_date, end_date)]
    display(mdf)
```

```python
#以圖表的方式呈現，看見2019-2020年公共網路使用的總流量趨勢
miaoli_wifi.plot(x='time', y='total_GB',title='19-20年的總使用量趨勢')
```

```python
#使用人次的趨勢
miaoli_wifi.plot(x='time', y='使用人次',title='19-20年的使用人次的變化')
```

```python
#用年度pd.Grouper分組
group_1Y=miaoli_wifi.groupby([pd.Grouper(key="time",freq='1Y')])
group_1Y.size()
```

```python
#知道19、20年總使用人數是多少?
group_1Y.使用人數.sum()
```

```python
#將使用人次排序，以高到低排序
miaoli_wifi_sort=miaoli_wifi.sort_values(by='使用人次',ascending=False)
miaoli_wifi_sort.loc[:,['time','total_GB','使用人次']]
```

```python
#列出使用平均人次
miaoli_wifi_mean=miaoli_wifi.使用人次.mean()
#找到低於平均使用人次的時間，並以高到低排序
miaoli_wifi_low=miaoli_wifi.loc[miaoli_wifi.使用人次 <=miaoli_wifi_mean]
miaoli_wifi_low.sort_values(by='使用人次',ascending=False)
```
