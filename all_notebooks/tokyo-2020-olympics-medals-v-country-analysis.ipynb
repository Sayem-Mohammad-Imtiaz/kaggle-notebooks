```python
import os, numpy as np, pandas as pd, matplotlib.pyplot as plt, seaborn as sns
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
coaches_df = pd.read_csv('/kaggle/input/tokyo-2020-olympics/coaches.csv')
technical_officials_df = pd.read_csv('/kaggle/input/tokyo-2020-olympics/technical_officials.csv')
athletes_df = pd.read_csv('/kaggle/input/tokyo-2020-olympics/athletes.csv')
medals_total_df = pd.read_csv('/kaggle/input/tokyo-2020-olympics/medals_total.csv')
medals_df = pd.read_csv('/kaggle/input/tokyo-2020-olympics/medals.csv')
print("Coaches dataframe shape",coaches_df.shape)
print("Technical officials dataframe shape",technical_officials_df.shape)
print("Athletes dataframe shape",athletes_df.shape)
print("Medals total dataframe shape",medals_total_df.shape)
print("Medals dataframe shape",medals_df.shape)
```

```python
print("Total medals won in 2020 Olympics",medals_total_df[['Total']].sum())
print("Gold medals won in 2020 Olympics",medals_total_df[['Gold Medal']].sum())
print("Silver medals won in 2020 Olympics",medals_total_df[['Silver Medal']].sum())
print("Bronze medals won in 2020 Olympics",medals_total_df[['Bronze Medal']].sum())
```

```python
# Checking value counts for each col, to understand the data;
def value_counts(dataframe):
    val_dict = {}
    for i in dataframe.columns: val_dict[i] = dataframe[[i]].value_counts()
    for i in val_dict.keys(): print("\n","\n-----------------------\n",val_dict[i],"\n-----------------------\n","\n\n")
    
value_counts(medals_total_df)
```

```python
# GOLD medals
sns.histplot(medals_total_df[['Gold Medal']], bins=medals_total_df.shape[0])
```

```python
# SILVER medals
sns.histplot(medals_total_df[['Silver Medal']], bins=medals_total_df.shape[0])
```

```python
# BRONZE medals
sns.histplot(medals_total_df[['Bronze Medal']], bins=medals_total_df.shape[0])
```

```python
# The first five countries in the medals_total_df have bagged more than half the Gold in 2020 Olympics, nothing new it has been that way since.
max_gold = medals_total_df[medals_total_df['Gold Medal']>int(medals_total_df[['Gold Medal']].max()/2)]
max_gold
```

```python
# GOLD medals
plt.rcdefaults()
fig, ax = plt.subplots()
ax.barh(max_gold['Country Code'].to_list(), max_gold['Gold Medal'].to_list(), align='center')
ax.set_xlabel('No. of medals')
ax.set_ylabel('Country Names')
ax.set_title('GOLD MEDALS')
plt.show()
```

```python
# SILVER medals
max_silver = medals_total_df[medals_total_df['Silver Medal']>int(medals_total_df[['Silver Medal']].max()/2)]

fig, ax = plt.subplots()
ax.barh(max_silver['Country Code'].to_list(), max_silver['Silver Medal'].to_list(), align='center')
ax.set_xlabel('No. of medals')
ax.set_ylabel('Country Names')
ax.set_title('SILVER MEDALS')
plt.show()
```

```python
# BRONZE medals
max_bronze = medals_total_df[medals_total_df['Bronze Medal']>int(medals_total_df[['Bronze Medal']].max()/2)]

fig, ax = plt.subplots()
ax.barh(max_bronze['Country Code'].to_list(), max_bronze['Bronze Medal'].to_list(), align='center')
ax.set_xlabel('No. of medals')
ax.set_ylabel('Country Names')
ax.set_title('BRONZE MEDALS')
plt.show()
```

```python
# TOTAL MEDALS:
temp_df = medals_total_df[medals_total_df['Total']>int(medals_total_df[['Total']].max()/2)]

fig, ax = plt.subplots()
ax.bar(temp_df['Country Code'].to_list(), temp_df['Total'].to_list(), align='center')
ax.set_xlabel('Country Names')
ax.set_ylabel('No. of medals')
ax.set_title('TOTAL MEDALS')
plt.show()
```
