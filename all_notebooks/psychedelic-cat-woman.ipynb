The project was approved at the 1973, 1975 and 1977 Biennials. During the event, Roberto stayed inside the House, distributing candles and taking visitors to the sensory experiences caused by mirrors. It was in the meantime that " Persona " was born ."

#Persona Disc - with the Songs to play.

"In the second half of the 1980s, Campadello started photographing his bar's customers playing Persona and developed a photographic technique to capture the images. At the same time, Roberto received the proposal from Grow (toy manufacturer) to produce the Persona Game on a large scale. Roberto goes to the United States and the “Persona Game” begins to be commercialized. The Game consisted of a Persona mirror, a pair of candles and candlestick and a Persona disc - with the songs to play. Failing to reach a satisfactory agreement with the company, Roberto stops the production of the Game and returns to Brazil. About a thousand Games were produced and marketed at that time. Today the Game and the Disc are regarded as relics - the Persona disc is the third rarest disc in the world."

```python
#Code by Asha Gutlapalli https://www.kaggle.com/ashagutlapalli/computer-vision-101-with-opencv/notebook

R = cv2.imread('../input/catsdogs/cats_and_dogs_filtered/train/cats/cat.10.jpg', cv2.IMREAD_COLOR)
R = cv2.resize(R, (224, 224))
H = cv2.imread('../input/beauty-detection-data-set/consolidated/beautiful/0004.jpg', cv2.IMREAD_COLOR)
H = cv2.resize(H, (224, 224))

G = R.copy()
guassian_pyramid_c = [G]
for i in range(6):
    G = cv2.pyrDown(G)
    guassian_pyramid_c.append(G)
 
     
G = H.copy()
guassian_pyramid_d = [G]
for i in range(6):
    G = cv2.pyrDown(G)
    guassian_pyramid_d.append(G)

laplacian_pyramid_c = [guassian_pyramid_c[5]]
for i in range(5, 0, -1):
    GE = cv2.pyrUp(guassian_pyramid_c[i])
    L = cv2.subtract(guassian_pyramid_c[i-1], GE)
    laplacian_pyramid_c.append(L)

laplacian_pyramid_d = [guassian_pyramid_d[5]]
for i in range(5,0,-1):
    guassian_expanded = cv2.pyrUp(guassian_pyramid_d[i])
    L = cv2.subtract(guassian_pyramid_d[i-1], guassian_expanded)
    laplacian_pyramid_d.append(L)

laplacian_joined = []
for lc,ld in zip(laplacian_pyramid_c, laplacian_pyramid_d):
    r, c, d = lc.shape
    lj = np.hstack((lc[:, 0 : int(c / 2)], ld[:, int(c / 2) :]))
    laplacian_joined.append(lj)

laplacian_reconstructed = laplacian_joined[0]
for i in range(1,6):
    laplacian_reconstructed = cv2.pyrUp(laplacian_reconstructed)
    laplacian_reconstructed = cv2.add(laplacian_reconstructed, laplacian_joined[i])

direct = np.hstack((R[ : , : int(c / 2)], H[ : , int(c / 2) : ]))

plt.figure(figsize=(30, 20))
plt.subplot(2,2,1).set_title('Cat', fontsize = 35); plt.axis('off')   
plt.imshow(cv2.cvtColor(R, cv2.COLOR_BGR2RGB))
plt.subplot(2,2,2).set_title('Woman', fontsize = 35); plt.axis('off')   
plt.imshow(cv2.cvtColor(H, cv2.COLOR_BGR2RGB))
plt.subplot(2,2,3).set_title('CatWoman Joining', fontsize = 35); plt.axis('off')   
plt.imshow(cv2.cvtColor(direct, cv2.COLOR_BGR2RGB))
plt.subplot(2,2,4).set_title('Pyramid Blending', fontsize = 35); plt.axis('off')   
plt.imshow(cv2.cvtColor(laplacian_reconstructed, cv2.COLOR_BGR2RGB))
plt.show()
```
