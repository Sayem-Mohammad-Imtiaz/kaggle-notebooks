### This dataset is comprised of weekly reports of eight human diseases throughout the 20th century.

##### - It has 700K+ rows and 7 columns. Actual counts and percentage per 100K people is included.
##### - This notebook gives a brief visualization of a single state's data.
##### - Diseases included are Hepatitis-A, Measles, Mumps, Pertussis, Polio, Rubella, Smallpox and Diptheria

- This data is from Project Tycho at healthdata.gov.
- The earliest records are from 1916, the most recent are from 2011. 
- Not all states reported for each year.
- It gives a good overview of the distribution of common diseases across the US.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
```

#### Let's load and take a look at the data.

```python
df = pd.read_csv('../input/diseases-in-the-us-1916-to-2011/Project_Tycho_Level_1_Data.csv', low_memory=False)
```

```python
# Check the shape, there should be 759,467 rows and 7 columns
df.shape
```

##### - No nulls, mostly object columns.
##### - The **epi_week** column is an int that represents the *year* and *week*. Each year has 01 to 52 weeks.
##### - The **epi_week** column and the **incidence_per_100000** column are numerical. The **cases** column is NOT numerical.
##### - We'll convert the **cases** column to a numerical value, so it's easier to plot.

```python
# Convert object column to numerical. We will lose any relavent strings here, so we'll leave the original column.
df['cases_numeric'] = pd.to_numeric(df['cases'], errors='coerce')
```

```python
# Verify the unique diseases in the dataset
df['disease'].unique()
```

#### Let's plot all the diseases reported for a single state. We'll choose California.

```python
# Grab all the results for CA
state_df = df[df['loc'] == 'CALIFORNIA']
```

```python
# Make a pairplot for California
sns.pairplot( hue='disease', data=state_df, height=4);
```

#### - Measles was huge compared to the other diseases .. Let's remove it and clean up the plots.

```python
# Remove Measles from the plot
state_df = state_df[state_df['disease'] != "MEASLES"]
sns.pairplot( hue='disease', data=state_df, height=4);
```

#### - Now we can see the distribution a little better.
#### - Let's see a more granular view.

```python
# Make a scatter plot
fig, ax = plt.subplots(figsize=(15, 6))
sns.scatterplot(x='epi_week', y='cases_numeric', hue='disease', data=state_df);
```

#### In CA Pertussis came, it went .. then came back it seems!

#### Rubella was big from the 80's until the 2000's.

##### - A lot of interesting insights can be extracted from this dataset with simple plotting.
