# Create a global map for Covid vaccination (Percentage of the population fully vaccinated)

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df=pd.read_csv ('../input/covid-world-vaccination-progress/country_vaccinations.csv', header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
print(df.columns)
fig=px.choropleth(df, locations = 'country', locationmode = 'country names', color = 'people_fully_vaccinated_per_hundred', hover_name = 'country', animation_frame = 'date', color_continuous_scale = 'Sunset', range_color = (0.0, 60.0))
fig.update_layout (title_text = "Percentage of the Country's Total Population that are Fully Vaccinated", title_x=0.40)
fig.show()

plotly.offline.plot(fig, filename = 'Full Vaccinations.html')
```

In this figure, we can visualize the the animation of percentage of Country total population those are vaccinated. From 26th December 2020, Germany began the fully-vaccination campaign. Until now, USA possess the highest percentage of people to be fully vaccinated, followed by some European countries. I have selected this topic to know about the trend of fully-vaccination around world.In some countries, vaccination becomes slow, e.g., in Japan (2.72 per hundred)

# Create a global map for Covid vaccination (Total Vaccinations Per Hundred)

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df=pd.read_csv ('../input/covid-world-vaccination-progress/country_vaccinations.csv', header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
print(df.columns)
fig=px.choropleth(df, locations = 'country', locationmode = 'country names', color = 'total_vaccinations_per_hundred', hover_name = 'country', animation_frame = 'date', color_continuous_scale = 'Blugrn', range_color = (0.0, 50.0))
fig.update_layout (title_text = "Total Vaccinations per Hundred in Different Countries", title_x=0.44)
fig.show()

plotly.offline.plot(fig, filename = 'Vaccinations per Hundred.html')
```

In this figure, we can visualize the the animation of Total Vaccinations per Hundred in Different Countries. From December 2020, Canada began the vaccination campaign followed by Russian, USA, UK and in some other EU countries. I have selected this topic to know about the trend of vaccination around world from the time being of its starting point of vaccination.
