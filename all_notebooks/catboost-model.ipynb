```python

import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
from sklearn.model_selection import train_test_split


fulltrain_df = pd.read_csv("/kaggle/input/heartbeat/mitbih_train.csv", header=None)
test_df = pd.read_csv("/kaggle/input/heartbeat/mitbih_test.csv", header=None)
fulltrain_df.to_numpy()
test_df.to_numpy()
```

```python
train_df, val_df = train_test_split(fulltrain_df, test_size = 0.33)
```

```python
x_train = train_df.loc[:,train_df.columns!=187]
y_train = train_df.loc[:,train_df.columns==187]

x_test = test_df.loc[:,test_df.columns!=187]
y_test = test_df.loc[:,test_df.columns==187]

x_val = val_df.loc[:,val_df.columns!=187]
y_val = val_df.loc[:,val_df.columns==187]
```

```python
from catboost import CatBoostClassifier

my_model = CatBoostClassifier(iterations=1200,learning_rate= 0.08,depth= 10,metric_period=5,bagging_temperature= 0.3, l2_leaf_reg=1)
```

```python
                        
my_model.fit(x_train,y_train,eval_set=(x_val,y_val),use_best_model=True, verbose= True)
```

```python

from sklearn.metrics import accuracy_score
y_pred  = my_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: %.2f%%" % (accuracy * 100.0))

from sklearn.metrics import mean_absolute_error
predictions = my_model.predict(x_test)
print("Mean Absolute Error : " + str(mean_absolute_error(predictions, y_test)))
```
