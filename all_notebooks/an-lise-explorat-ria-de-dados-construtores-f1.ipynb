```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../input/construtores-formula-1/ConstrutoresFormula1.csv')
df.head()
```

```python
plt.figure(figsize=(20,8))
df.groupby('Equipe')['Margem'].max().sort_values().plot.barh()
plt.title('Maior margem de pontos por equipe campeã')

```

```python
menores_margens = df[df['Margem'] <=8 ]
menores_margens
```

```python
df['Margem'].nlargest(10)
```

```python
maiores_margens = df[df['Margem'] >= 134]
maiores_margens
```

```python
plt.figure(figsize=(20,8))
sns.scatterplot(data=df,x='Temporada',y='Margem')

```

```python
sns.lineplot(data=df,x='Temporada',y='Margem')
```

```python
df.tail(8)
```

Nota-se que nas temporadas mais recentes as margens que as equipes campeãs têm sobre as vice-campeãs são maiores do que as temporadas iniciais da categoria. Desde o início da chamada "era híbrida", em 2014, em que os motores possuem capacidade de recuperação de energia, a equipe Mercedes é dominante absoluta, com a menor margem sendo de 84 pontos, o que no sistema de pontuação atual significa 3 vitórias em corridas e mais um 5° lugar. Impressionante!

```python
df.groupby('Equipe')['Podios'].max().sort_values().plot.bar(ylabel='Máximo de pódios por temporada')
```

```python
df.groupby('Equipe')['Poles'].max().sort_values().plot.bar(ylabel='Máximo de poles por temporada')
```

```python
df.groupby('Motor')['Voltas rapidas'].max().sort_values().plot.bar(ylabel='Máximo de voltas rápidas por temporada')
```

A equipe de Fórmula 1 Mercedes foi fundada em 2010 mas já aparece entre as recordistas em várias estatísticas como as vistas acima contra equipes bem mais antigas. É impressionante o que essa equipe vem fazendo no esporte em tão pouco tempo.
