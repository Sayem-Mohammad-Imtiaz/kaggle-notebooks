```python
sns.heatmap(df.loc[:,[True  if col not in ['SmartCity_Index_relative_Edmonton','SmartCity_Index'] else False for col in df.columns]].corr(),annot=True)
```

The above heat map is drawn so as to find the correlation between the features.
    + Smart City Index is removed as it is a dependent feature and defines the entire city ranking based on the other individual features.

+ Based on the heatmap correlation matrix, 
    + Smart Living and Smart government are correlated 
    + Smart Economy and Smart Environment are correlated
    + Smart People and Smart mobility are correlated


```python
sns.heatmap(df.loc[:,df.columns!='SmartCity_Index_relative_Edmonton'].corr(),annot=True)
```

Now, If we look at the Correlation for Smart city index.
+ Smart living is highly correlated and Smart mobility is least correlated(but it is significant).
+ Smart Environment, Smart government, smart people also contribute somewhat to smartcity_index.
