```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1=pd.read_csv('../input/netflix2020/DataNetflixRevenue2020_V2.csv') 
df1.head() #DataNetflixRevenue2020
```

```python
df2=pd.read_csv('../input/netflix2020/DataNetflixSubscriber2020_V2.csv')
df2.head() #DataNetflixSubscriber2020
```

```python
df3=pd.read_csv('../input/netflix2020/NetflixSubscribersbyCountryfrom2018toQ2_2020.csv')
df3.head() #NetflixSubscribersbyCountryfrom2018toQ2_2020
```

```python
df4=pd.read_csv('../input/netflix2020/NetflixsRevenue2018toQ2_2020.csv')
df4.head() #Netflix'sRevenue2018toQ2_2020
```

```python
print(df1.shape)
print(df2.shape)
print(df3.shape)
print(df4.shape)
```

```python
rev_sub=pd.concat([df1,df2],axis=1)
rev_sub=rev_sub.iloc[:,2:]
rev_sub.head()
```

```python
rev_sub.info()
```

No null entries

```python
rev_sub.Years.unique()
```

```python
plt.bar(rev_sub.Years,rev_sub.Revenue)
plt.xticks(rotation=45)
plt.xlabel('Year')
plt.ylabel('Dollars ($)')
plt.legend()
plt.show()
```

Revenue has increased with time

```python
plt.bar(rev_sub.Years,rev_sub.Subscribers,color='g')
plt.xticks(rotation=45)
plt.xlabel('Year')
plt.ylabel('Dollars ($)')
plt.legend()
plt.show()
```

Subscribers have increased with time

```python
rev_sub.groupby(['Area']).mean()
```

```python
df1.groupby(['Area']).mean().plot(kind='bar')
df2.groupby(['Area']).mean().plot(kind='bar')
plt.grid()
plt.xticks(rotation=90)
plt.show()
```
