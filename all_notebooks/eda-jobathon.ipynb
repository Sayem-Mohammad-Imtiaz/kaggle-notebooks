# Key Points:

1. It's a imbalanced problem, so you need to apply some technique either SMOTENC,Over/UnderSampling or imblearn algos like BalancedRandomForest

2. Imputing 12% values with mode might or might not create problem, so test the model for both.

3. Test Data also has missing values for credit, change them using mode of training data and not test (Important).

4. Avg_Account_Balance has many outliers, choose a algo accordingly. Else, maybe you can change outliers with median value (You have to check accuarcy of model to decide that)

5. Accuracy won't be a good metric here. Precision Recall curve is robust to imbalanced datasets. So, use that while evaluating and not ROC-AUC as it's good for balanced classes.

6. When you're done, scale the data , it would be make algo run comparatively fast. Make sure to fit only train data and use the result to transfrom test data.

7. Be careful while scaling and imputing values, don't leak the test data to the model.
