<h1 style="text-align:center;font-weight: bold">About the Dataset</h1>
    <h3 style="text-align:left;font-weight: bold">A data frame with 322 observations of major league players on the following 20 variables.
    </h3>
<ul>
<li>AtBat: Number of times at bat in 1986</li>
<li>Hits: Number of hits in 1986</li>
<li>HmRun: Number of home runs in 1986</li>
<li>Runs: Number of runs in 1986</li>
<li>RBI: Number of runs batted in in 1986</li>
<li>Walks: Number of walks in 1986</li>
<li>Years: Number of years in the major leagues</li>
<li>CAtBat: Number of times at bat during his career</li>
<li>CHits: Number of hits during his career</li>
    <li>CHmRun: Number of home runs during his career</li>
    <li>CRuns: Number of runs during his career</li>
    <li>CRBI: Number of runs batted in during his career</li>
    <li>CWalks: Number of walks during his career</li>
    <li>League: A factor with levels A and N indicating player's league at the end of 1986</li>
    <li>Division: A factor with levels E and W indicating player's division at the end of 1986</li>
    <li>PutOuts: Number of put outs in 1986</li>
    <li>Assists: Number of assists in 1986</li>
    <li>Errors: Number of errors in 1986</li>
    <li>Salary: 1987 annual salary on opening day in thousands of dollars</li>
    <li>NewLeague: A factor with levels A and N indicating player's league at the beginning of 1987</li>
</ul>

<h2 style="text-align:center;font-weight: bold">Linear Regression</h2></div>



```python
params_lin ={'fit_intercept': True, 'positive': True}
regressor = Regression(predictor = 'lin', params=params_lin, pca='y', pca_kernel = 'rbf', n_components_pca =3)
regressor.fit(features, labels)
result = regressor.result()
accuracy_scores[result['Regressor']] = result['Accuracy']
RMSE[result['Regressor']] = result['RMSE']
```
