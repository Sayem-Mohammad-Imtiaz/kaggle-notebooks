# CODE STARTS FROM HERE

```python
from fastai import *
from fastai.vision import *
```

```python
dest = '../outputs/Dataset/'
```

```python
np.random.seed(2)
pat = r'/([^/]+)_\d+.jpg$'
```

```python
fnames = get_image_files(dest)
fnames[:5]
```

```python
data = ImageDataBunch.from_name_re(dest, fnames, pat ,ds_tfms = get_transforms(),size = 224)
data.normalize(imagenet_stats)
```

```python
data.show_batch(rows = 3)
```

```python
data.c
```

```python
data.classes
```

```python
learner=create_cnn(data,models.,metrics=[accuracy])
```

```python
learner.fit_one_cycle(5)
```

```python
learner.unfreeze() # must be done before calling lr_find
learner.lr_find()
learner.recorder.plot()
```

```python
learner.fit_one_cycle(5, max_lr=slice(3e-5, 3e-4))
```

```python
learner.model_dir = "/tmp/model/"
learner.save("resnet101-stage-1")
```

```python
learner.unfreeze() # must be done before calling lr_find
learner.lr_find()
learner.recorder.plot()
```

```python
learner.fit_one_cycle(8, max_lr=slice(3e-6, 3e-4))
```

```python
learner.model_dir = "/kaggle/working"
learner.save('resnet101-stage-2-78')
```
