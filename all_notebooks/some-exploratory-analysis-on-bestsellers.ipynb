# Data Visualizations on Authors

### Top 5 authors of all time?

```python
df.Author.value_counts().head()

df.Author.value_counts().head().plot(kind='bar', figsize=(9,6), title='Top 5 bestselling authors')
```

### Top 3 authors in each year?

```python
temp = df.value_counts(['Year', 'Author']).sort_index().reset_index()
```

```python
temp = temp.rename(columns={0: 'count'}).sort_values(by=['Year', 'count'], ascending=[True, False])
temp.head()
```

```python
for y in temp.Year.unique():
    temp_y = temp[temp.Year == y]
    print('Top 3 authors of {}: {}'.format(y, ', '.join(temp_y.Author.head(3))))
```

### Top 5 authors for each genre?

```python
temp = df.value_counts(['Genre', 'Author']).sort_index().reset_index()
temp = temp.rename(columns={0: 'count'}).sort_values(by=['Genre', 'count'], ascending=[True, False])
temp.head()
```

```python
for g in temp.Genre.unique():
    temp_g = temp[temp.Genre == g]
    print('Top 5 {} authors: {}'.format(g, ', '.join(temp_g.Author.head(5))))
```

```python
for g in temp.Genre.unique():
    temp[temp.Genre == g].head().plot(kind='bar', x='Author', y='count', 
                                      figsize=(9,6), title='Top 5 {} Authors'.format(g))
```

### Top 5 highest average user ratings for each author?

```python
df_author = df.groupby('Author').mean()
```

```python
df_author.sort_values(['User Rating', 'Reviews'], ascending=False).head()
```
