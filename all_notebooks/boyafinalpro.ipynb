##  <span class="mark" style="color:#000">Explore the causes of  employee attrition  for enterprises</span>

### What are the Department & MonthlyIncome  impact on  Attrition?

```python
dma=df[['Department','MonthlyIncome','Attrition']]
plt.figure(figsize=(8,5)) 
ax = sns.boxplot(x="Department", y="MonthlyIncome", hue="Attrition",data=dma,linewidth=1.5,order=["Research & Development", "Sales",'Human Resources'])
plt.show()
```

___conclusion:___ It can be seen that the people who choose to quit work are mainly in the Sales Department. For them, the monthly salary has little impact, while for the people who quit work in the other two departments, it can be seen that the monthly salary is relatively low, which is the reason for their resignation.It is suggested that the company can appropriately raise the salaries of these two departments.
__(๑•̀ㅂ•́)و✧__

### What are the JobRole & Age impact on Attrition? 

```python
plt.figure(figsize=(8,6)) 
jaa=df[['JobRole','Age','Attrition']]
ax = sns.boxplot(x="Age", y="JobRole", hue="Attrition",data=jaa,linewidth=1.5, palette="Set1",orient="h")
plt.show()
```

___conclusion:___ It can be seen the staff quiting work of whoes job function is sales representative or human resources,they are younger. While most of the dimission personnel of manager and research director are older, and the age of the dimission personnel of the remaining departments is in the middle.So it is recommended that companies pay more attention to what is happening to young people in these two sectors (sales & human)and to what is happening to older people in (manager & research director).╰(*°▽°*)╯

###   What are the MonthlyIncome & Gender & MaritalStatus impact on Attrition?

```python
amg=df[['Attrition','MonthlyIncome','Gender','MaritalStatus']]
ax= sns.catplot(x="MaritalStatus", y="MonthlyIncome",hue="Attrition", col="Gender",data=amg, kind="box",height=6, aspect=.8);
```

___conclusion___:It can be seen that both men and women, low salary is a major factor affecting the departure,and the number of married people who left the workforce was the highest. Most intriguingly, divorced women are among the least likely to leave, and companies can be tempted to hire them if they want to retain talent.┑(￣Д ￣)┍

### What are the MonthlyIncome & OverTime impact on Attrition? (Explore only the low-paid)

```python
ax = sns.violinplot(x="OverTime", y="MonthlyIncome",hue="Attrition",split=True,data=df[df.MonthlyIncome < 10000],scale="width", palette="muted")
```

___conclusion___: It can be seen that whether or not overtime work has little impact on the dimission. What is interesting is whether overtime work will affect the dimission of this group of people with a monthly income of about 6000. Σ( ° △ °|||)︴

### What are the NumCompaniesWorked & BusinessTravel impact on Attrition?

```python
bt = df[["BusinessTravel","Attrition"]]
bt=bt.groupby(["BusinessTravel","Attrition"]).size().reset_index()
#bt=DataFrame(bt,index=['Non-Travel-N', 'Non-Travel-Y', 'Travel_Frequently-N', 'Travel_Frequently-Y','Travel_Rarely-N','Travel_Rarely-Y'],columns=("BusinessTravel","Attrition",'count'))
bt.rename(columns={0:'Count'},index={0:'Non-Travel-N',1:'Non-Travel-Y', 2:'Travel_Frequently-N', 3:'Travel_Frequently-Y',4:'Travel_Rarely-N',5:'Travel_Rarely-Y'},inplace=True)
bt
```

```python
bt.plot(kind='pie',y='Count',figsize=(6,6),title='travel&attrition-analysis',fontsize=12,explode=[0,0,0,0.2,0,0],shadow=True,startangle=60)
```

```python
ax = sns.countplot(x="BusinessTravel", hue="Attrition", data=df,palette="bright")
```

```python
ax = sns.countplot(x="NumCompaniesWorked", hue="Attrition", data=df,palette="autumn")
```

___conclusion___: Since people who rarely travel account for a large proportion, there is a phenomenon of more attrition. The actual travel has little relationship with the dimission. It is worth noting that the number of dimission accounts for one third of the frequent travelers, which accounts for a relatively large number, indicating that frequent travelers are prone to leave.Similarly, people who have worked for 5-9 companies are more likely to leave.♪(^∇^*)
