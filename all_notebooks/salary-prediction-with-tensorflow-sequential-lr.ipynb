I predicted salary with the Sequential and Linear regressions

```python
salary_data = pd.read_csv('/kaggle/input/salary/Salary.csv')
```

```python
salary_data.head()
```

```python
#visualise the data
plt.figure(figsize=(16,8))
plt.title('Salary vs Experince')
plt.xlabel('Experince (Years)')
plt.ylabel('Salaries - Annualy - $(USD)')

plt.plot(salary_data)
```

```python
#since data is clear we won't do data clearning stuffs, this a regression problem
X = salary_data.iloc[:, 0]
y= salary_data.iloc[:,1]

#split the data
X_train,X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
def get_model():
    model = Sequential([
        Dense(10000, input_shape = (1,), activation='relu'),
        Dense(5000, activation='relu'),
        Dense(2500, activation='relu'),
        Dense(1750, activation='relu'),
        Dense(875, activation='relu'),
        Dense(476, activation='relu'),
        Dense(238, activation='relu'),
        Dense(119, activation='relu'),
        Dense(60, activation='relu'),
        Dense(30, activation='relu'),
        Dense(15, activation='relu'),
        Dense(1)
    ])
    model.compile(
        loss='mse',
        optimizer='adam'
    )
    return model
get_model().summary()
```

```python
es_cb = EarlyStopping(monitor='val_loss', patience=1000)
model = get_model()

preds_on_intrained = model.predict(X_test)

history = model.fit(
    X_train, y_train,
    validation_data = (X_test, y_test),
    epochs = 1000,
    callbacks = [es_cb]
)
```

```python
preds_on_trained = model.predict(X_test)
```

```python
preds_on_trained
```

```python
rmse = np.sqrt( np.mean(preds_on_trained - np.array(y_test)) ** 2)
print('rsme : {}'.format(rmse))

#modele not has some big Root Mean Square Error (RMSE) but still given bit close prediction. I am personaly not like to deal with big RMSE
```

```python
preds_on_trained = model.predict(X)
```

```python
salary_data['Predictions'] = preds_on_trained
```

```python
salary_data['Predictions'] = salary_data['Predictions'].apply(lambda x: math.ceil(x))
```

```python
plt.figure(figsize=(16,8))
plt.title('Salary actual value and Prediction')
plt.xlabel('Number of Experince (Years)')
plt.ylabel('Salary - $(USD)')
plt.plot(salary_data['Salary'])
plt.plot(salary_data['Predictions'])
plt.legend(['Actual Salary', 'Predicted Salary by Model'], loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.title('Salary actual value and Prediction')
plt.xlabel('Number of Experince (Years)')
plt.ylabel('Salary - $(USD)')
plt.scatter(X_train,y_train,  color = 'red')
plt.plot(X_train,model.predict(X_train),  color = 'blue')
plt.legend(['Actual Salary', 'Predicted Salary by Model'], loc='upper left')
plt.show()
```
