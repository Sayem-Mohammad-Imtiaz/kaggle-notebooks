```python
#Import necessary libraries, read csv files and store to a dataframe (df):

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

#Load GDP data for each region
df = pd.read_csv('../input/country-regional-and-world-gdp/gdp_csv.csv')

#Load list of countries:
df2 = pd.read_csv('../input/countries/countries of the world.csv')
```

```python
#Look at the columns and values present for the GDP dataframe
df.head()
```

```python
#Look at the columns and values present for the countries dataframe
df2.tail()
```

It looks like regions are included in the df dataset. Lets use another dataset to ensure we only complete analysis on countries, not regions. This is done by comparing the respective columns in both datasets.

```python
df['Country Name'].unique()
```

```python
countries = list(df2['Country'].unique())
countries
```

The above shows that there is a space at the end of every string, now we need to remove the space and then use the pandas isin method to compare the values for countries in the two datasbases.

```python
#Using list comprehension to remove blank space:
countries = [country.replace(' ','') for country in countries]
countries
```

```python
#Compare list of countries from df2 to country column of df. Where countries in both databases are assigned a value of true.
new = df['Country Name'].isin(countries)
#Create new dataframe which reflects data only from countries:
df_new = df[new]
#Reset index to start from 0
df_new.reset_index(drop = True, inplace = True)
#Normalise the GDP value:
df_new['Value'] = (df_new['Value'])/max(df_new['Value'])
df_new.head()
```

However, unfortunately, not all countries in countries dataset is in the original GDP data set. Therfore, it is necessary to drop rows where countries are not present in the original data set.

```python
#Start by finding out which countries are included:
countries_df = pd.DataFrame(countries)

included = countries_df.isin(df['Country Name'].values.tolist())

countries_included = countries_df[included]
countries_included.head()
```

```python
#As seen from above, rows with not a number values, are the rows of countries excluded from the original dataset.
rows_with_nan = []
for index, row in countries_included.iterrows():
    is_nan_series = row.isnull()
    if is_nan_series.any():
        rows_with_nan.append(index)
#rows_with_nan represents the index of missing countries from the countries_df
print(rows_with_nan)
```

```python
#Create dataframe with the missing countries
countries_excluded = countries_df.iloc[rows_with_nan]
#Convert ot list to easily drop from df_new
countries_excluded.values.tolist()
```

```python
df_new.drop(countries_excluded, axis = 0, inplace = True)
df_new
```

Country code does not add any value to analysis, therefore the column can be dropped

```python
df_new.drop(['Country Code'], axis = 1, inplace = True)
df_new.head()
```

```python
import matplotlib.pyplot as plt

afghanistan = df_new[df_new['Country Name'] == 'Afghanistan']

plt.plot(afghanistan['Year'], afghanistan['Value'])
plt.xlabel('Year')
plt.ylabel('GDP')
plt.show()
```
