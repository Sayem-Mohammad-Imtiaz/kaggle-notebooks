```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams['figure.figsize'] = 10,8
import warnings
warnings.filterwarnings('ignore')
```

```python
dataset_2015 = pd.read_csv('../input/2015.csv')
```

```python
dataset_2015.describe()
```

# 2015 Data

## Happiness vs Health(Life Expectancy)


```python
vis1 = plt.scatter(x = dataset_2015['Happiness Score'], y = dataset_2015['Health (Life Expectancy)']);
plt.xlabel('happiness Score');
plt.ylabel('Health (Life Expectancy)');
```

## Happiness score vs Freedom


```python
vis2 = plt.scatter(x = dataset_2015['Happiness Score'], y = dataset_2015['Freedom']);
plt.xlabel('happiness Score');
plt.ylabel('Freedom');
```

## Happines vs trust

```python
vis3 = sns.lmplot(data = dataset_2015, x = "Happiness Rank", y = "Trust (Government Corruption)", hue = "Region", \
                  fit_reg = False,  scatter_kws = {"s": 100},size = 10);
```

## Trust vs region

```python
vis4 = sns.barplot(data = dataset_2015, x = "Trust (Government Corruption)", y = "Region",)
```

As we can see that Australia and New Zealand has the highest trust level followed by North America and then by Westen Europe. The least level belongs to Southern Asia maybe because of high level of corruption.

### 2015 vs 2016

```python
fig, axs = plt.subplots(ncols=2,sharex=True,sharey = True)
sns.barplot(data = dataset_2015, x = "Trust (Government Corruption)", y = "Region", ax=axs[0]);
sns.barplot(data = dataset_2016, x = "Trust (Government Corruption)", y = "Region", ax=axs[1]);
```

As we can see here. The Austalians have highest trust level on government and it increased futher.
North American trust in their Government decreased a bit. Latin America and Caribbean shows a lot of change.
With Southern Asia still at the last and negligivle change.

```python
fig, axs = plt.subplots(ncols=2,sharex=True,sharey = True)
sns.barplot(data = dataset_2015, x = "Freedom", y = "Region", ax=axs[0]);
sns.barplot(data = dataset_2016, x = "Freedom", y = "Region", ax=axs[1]);
```

Freedom in North America increased a lot whereas in Australia it decreased. The huge negative change is shown by Middle East and Northern Africa along with Central and Eastern Europe.

```python
fig, axs = plt.subplots(ncols=2,sharex=True,sharey = True)
sns.barplot(data = dataset_2015, x = "Health (Life Expectancy)", y = "Region", ax=axs[0]);
sns.barplot(data = dataset_2016, x = "Health (Life Expectancy)", y = "Region", ax=axs[1]);
```

The Health (Life Expectancy) decreased in all the regions and there is a need to look into it as it impacts the happiness index a lot.
