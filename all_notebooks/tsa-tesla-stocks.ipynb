```python
prophet = fbprophet.Prophet(changepoint_prior_scale=0.05, n_changepoints=10)
prophet.fit(train)
pred_model = prophet.make_future_dataframe(periods=192, freq='D')
pred = prophet.predict(pred_model)
prophet.plot(pred)
```

```python
prophet.plot_components(pred)
```
