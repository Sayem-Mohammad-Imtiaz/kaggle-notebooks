```python
# independent variables
x = df.drop(['quality'],axis = 1)
```

```python
x
```

```python
# depenedent variable
y = df['quality']
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y, test_size=0.1)
```

```python
# training the model
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
predict_value = np.array([int(i) for i in model.predict(X_test)])
```

```python
predict_value
```

```python
y_test
```

```python
model.intercept_
```

```python
model.coef_
```

```python
model.score(X_test, y_test)
```

```python
# implementing y = mx + c
y_pred = (X_test.dot(model.coef_))+model.intercept_
y_pred
```

```python
r2_score(y_test,model.predict(X_test))
```

```python
r2_score(y_test,y_pred)
```

```python
model.score(X_test,y_test)
```
