### Melihat data dari `database.csv`

```python
df = pd.read_csv('/kaggle/input/spacex-missions/database.csv')
df.head()
```

### Melihat semua nama dan tipe data kolom database.csv

```python
df.info()
```

### Kemungkinan sukses peluncuran roket

```python
keberhasilan = df.groupby(['Mission Outcome'])['Flight Number'].count()

plt.bar(keberhasilan.index, keberhasilan)
plt.ylabel("Jumlah")
plt.title("Keberhasilan")
plt.show()
```

### Jenis pelanggan dari roket yang akan diluncurkan oleh SpaceX

```python
jenis = df.groupby(['Customer Type'])['Flight Number'].count()

plt.bar(jenis.index, jenis)
plt.ylabel('Jumlah')
plt.show()
```

### Nama pelanggan SpaceX yang meluncurkan roket

```python
launchName = df.groupby(['Customer Name'])['Flight Number'].count()

plt.barh(launchName.index, launchName)
plt.xlabel('Jumlah')
plt.show()
```

### Berat barang yang dibawa oleh roket (kalau membawa)

```python
berat_rata = df['Payload Mass (kg)'].dropna().mean()
berat_min = df['Payload Mass (kg)'].dropna().min()
berat_max = df['Payload Mass (kg)'].dropna().max()

print(f"Rata-rata : {berat_rata}")
print(f"Minimal : {berat_min}")
print(f"Maksimal : {berat_max}")
```

### Presentase tempat peluncuran

```python
tempat = df.groupby(['Launch Site'])['Flight Number'].count() / df['Flight Number'].count() * 100

plt.barh(tempat.index, tempat)
plt.ylabel('Persentase')
plt.show()
```

### Jenis roket yang diluncurkan

```python
jenisRoket = df.groupby(['Vehicle Type'])['Flight Number'].count()

plt.barh(jenisRoket.index, jenisRoket)
plt.ylabel('Persentase')
plt.show()
```

### Tipe barang yang dibawa roket

```python
df['bawaan'] = df['Payload Type'].fillna('Tidak Ada')
bawaan = df.groupby(['bawaan'])['Flight Number'].count()

plt.barh(bawaan.index, bawaan)
plt.xlabel('Jumlah')
plt.show()
```

### Tipe orbit satelit

```python
df['orbit'] = df['Payload Orbit'].fillna('Tidak ada satelit')
orbit = df.groupby(['orbit'])['Flight Number'].count()

plt.barh(orbit.index, orbit)
plt.xlabel('Jumlah')
plt.show()
```
