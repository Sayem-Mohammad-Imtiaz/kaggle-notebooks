### Step 2. Preprocessing

```python
df[df['topic'].isin(['Data-Scienc','machine-Learning','Open-CV','Computer-Vision','GAN','variational-encoder'])]
```

```python
ds_df = df[df['topic'].isin(['Data-Scienc','machine-Learning','Open-CV','Computer-Vision','GAN','variational-encoder'])]
```

```python
#exclude repos - with low stars
ds_df[ds_df['star'].str.contains('k') == True]
```

```python
# save new df
ds_df = ds_df[ds_df['star'].str.contains('k') == True]
```

```python
#exclude repos - with low forks
ds_df[ds_df['fork'].str.contains('k') == True]
```

```python
ds_df = ds_df[ds_df['fork'].str.contains('k') == True]
```

```python
# transform dtype - column : pull_requests 
ds_df['pull_requests'] = ds_df['pull_requests'].str.replace(",","")
ds_df['pull_requests'] = ds_df['pull_requests'].astype(int)
```

```python
ds_df['pull_requests']
```

```python
#exclude repos - with low pull_requests
ds_df[ds_df['pull_requests'] > 100]
```

```python
top11_ds = ds_df[ds_df['pull_requests'] > 100]
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
plt.style.use(['seaborn-white'])
```

```python
plt.figure(figsize=(20,6))
sns.barplot(data=top11_ds, x='name',y='pull_requests')
```

```python
# transform dtype - column : star
ds_df['star'] = ds_df['star'].str.replace(".","")
ds_df['star'] = ds_df['star'].str.replace("k","1000")
ds_df['star'] = ds_df['star'].astype(int)
```

```python
# transform dtype - column : star
ds_df['fork'] = ds_df['fork'].str.replace(".","")
ds_df['fork'] = ds_df['fork'].str.replace("k","1000")
ds_df['fork'] = ds_df['fork'].astype(int)
```

```python
# transform dtype - column : star
ds_df['watch'] = ds_df['watch'].str.replace(".","")
ds_df['watch'] = ds_df['watch'].str.replace("k","1000")
ds_df['watch'] = ds_df['watch'].astype(int)
```

```python
top11_ds = ds_df[ds_df['pull_requests'] > 100]
```
