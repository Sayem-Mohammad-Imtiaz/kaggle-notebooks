```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression

plt.style.use('seaborn')
```

```python
df = pd.read_csv('/kaggle/input/covid19-indonesia/covid_19_indonesia_time_series_all.csv')
df.head()
```

```python
data = df[df['Location'] != 'Indonesia'] # Remove location = indonesia from candidate
data
```

```python
dataJakarta = data[data['Location'] == 'DKI Jakarta']
dataJakarta
```

```python
dataNovember = dataJakarta[pd.DatetimeIndex(dataJakarta['Date']).month == 11]
dataNovember
```

```python
tanggal = pd.DatetimeIndex(dataNovember['Date']).day.values
newCase = dataNovember['New Cases'].values

x_train = tanggal[:,np.newaxis]
y_train = newCase

lm = LinearRegression()
lm.fit(x_train, y_train)
```

```python
pb  = lm.predict(x_train)
dfc = pd.DataFrame({'x': tanggal,'y': pb})

plt.scatter(tanggal, newCase)
plt.plot(dfc['x'], dfc['y'], color='red', linewidth=2)
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kasus')
plt.show()
```
