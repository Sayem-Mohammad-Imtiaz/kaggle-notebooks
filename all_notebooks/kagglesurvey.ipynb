```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../input/datascience-survey-on-kaggle/kagglesurvey.csv')
df.head()
```

```python
df = df.set_index(df.Respondent)
df.head(2)
```

```python
df=df.drop('Respondent', axis=1)
```

```python
df.LanguageRecommendationSelect.unique()
```

```python
lr = df.LanguageRecommendationSelect.value_counts()
lrs = df.LanguageRecommendationSelect.value_counts().to_dict()


fig, ax = plt.subplots(2,1,figsize=(15,15))
sns.barplot(lr.values, lr.index, color='y', ax=ax[0])
ax[1].pie(lrs.values(), labels=lrs.keys(), autopct='%1.1f%%')
plt.show()
```

```python
ei = df.EmployerIndustry.value_counts().to_dict()

plt.figure(figsize=(8,8))
plt.pie(ei.values(), labels=ei.keys(), autopct='%1.1f%%')
plt.show()
```

```python
wa=df.WorkAlgorithmsSelect.value_counts().head(10)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(wa.values, wa.index, color='r')
```
