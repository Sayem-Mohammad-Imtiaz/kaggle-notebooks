```python
%matplotlib inline

# Load the dataset
in_file = '/kaggle/input/pakistan-car-reviews/Pakistan Car Reviews.csv'

full_data = pd.read_csv(in_file,encoding='mac_roman')

full_data.head()
```

```python
isnull = full_data.isnull().sum()
isnull.head()
```

```python
replacement = {
    "-rating": "",
}

full_data['Style']=full_data['Style'].replace(replacement, regex=True)
full_data['Comfort']=full_data['Style'].replace(replacement, regex=True)
full_data['Fuel Economy']=full_data['Style'].replace(replacement, regex=True)
full_data['Performance']=full_data['Style'].replace(replacement, regex=True)
full_data['Value for Money']=full_data['Style'].replace(replacement, regex=True)
full_data
```

```python
full_data.dtypes
```

```python
full_data['Comfort']  = full_data['Comfort'].astype('int')
full_data['Fuel Economy']  = full_data['Fuel Economy'].astype('int')
full_data['Performance']  = full_data['Performance'].astype('int')
full_data['Value for Money']  = full_data['Value for Money'].astype('int')
full_data['Style']  = full_data['Style'].astype('int')
full_data.dtypes
```

```python
full_data['totalratings'] = full_data['Value for Money'] + full_data['Performance'] + full_data['Comfort'] + full_data['Style'] + full_data['Fuel Economy']
```

```python
full_data
```

```python
full_data=full_data[~full_data['Review'].str.contains('Do you find this review helpful')]
full_data
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
full_data=full_data.reset_index(drop=True)
full_data
```

```python
full_data['positivity_in_review'] = 0.0
for index, row in full_data.iterrows():
    score=sia.polarity_scores(row['Review'])
    full_data.positivity_in_review.iloc[index] = score['pos']
full_data
```

```python
max_rating_df=full_data[full_data["totalratings"]==25]
```

```python
max_rating_df=max_rating_df.reset_index(drop=True)
max_rating_df
```
