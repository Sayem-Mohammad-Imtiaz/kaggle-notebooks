## Covid 19 Deaths by Country
#### taken from dunderdata tutorials to show the "bar_chart_race" function

```python
!pip install bar_chart_race

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from IPython.display import HTML,display
import bar_chart_race as bcr
import warnings

warnings.filterwarnings("ignore")

df = pd.read_csv('/kaggle/input/covid-19/data/countries-aggregated.csv')
df_chart = df.pivot(index='Date', columns='Country', values='Deaths')
df_chart = df_chart.fillna(df_chart.mean())
```

```python
df_chart_simplified = df_chart.iloc[::2]

bcr.bar_chart_race(df=df_chart_simplified,
                              filename=None,
                              orientation='h',
                              sort='desc',
                              n_bars=10,
                              label_bars=True,
                              steps_per_period=10,
                              period_length=500,
                              figsize=(3, 5),
                              cmap='dark24',
                              title='Covid-19 Deaths',
                              bar_label_size=5,
                              tick_label_size=5,
                              fig=None)

#thank you Ted , https://github.com/dexplo/bar_chart_race just wanted to show you this easy function 
```
