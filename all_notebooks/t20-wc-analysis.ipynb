# Converting into Numerical columns

```python
# Convert B,4s,6s,SR columns to numerical columns
df.info()
```

```python
# function to convert str to int 
def num(x):
    if x != '-':
        return int(x)
    else:
        return 0
    
# function to convert str to float   
def numfloat(x):
    if x != '-':
        return float(x)
    else:
        return 0
```

```python
# coverting to numerical columns
df['6s']= df['6s'].apply(num)
df['4s']= df['4s'].apply(num)
df['SR']= df['SR'].apply(numfloat)
```

```python
df['B']= df['B'].apply(num)
```

```python
# COnversion completed
df.info()
```

```python
# function for finding name of wicket taken by bowler from wicket column
import re
def bowler(t):
    if 'b' in t:
        i = re.search('b ',t)
        try:
            ind = i.end()
            return (t[ind::])
        except:
            pass
    
    else:
        return t
```

```python
# function for finding name of catch taken by fielder from wicket column

def catch(t):
    if 'c & b' in t:
        i = re.search('c & b',t)
        return t[i.end()+1::]
        
    
    elif 'c' in t:
        try:
            c = re.search('c ',t)
            b = re.search('b ',t)
            return t[c.end():b.start()-1]
        except:
            pass
    else:
        return '-'
```

```python
# function for finding batsman runout or not from wicket column

def runout(t):
    if 'run out' in t:
        return 'Run Out'
    else:
        return '-'
    
```

```python
# function for finding team won from Winner column
def won(t):
    if 'won' in t:
        i = re.search('won',t)
        return t[:i.start()-1]
    else:
        return t
```

```python
# apply function to winner column and create new won column
df['Won'] = df['Winner'].apply(won)
```

```python
# apply function to Wicket column and create new runout column
df['Run Out'] = df['Wicket'].apply(runout)
```

```python
# apply function to Wicket column and create new catched taken by column
df['Catched By'] = df['Wicket'].apply(catch)
```

```python
# apply function to Wicket column and create new bowler column
df['Bowler'] = df['Wicket'].apply(bowler)
```

```python
# won column has match tied and won by that team records find won team from it
def winner(t):
    if 'Match tied' in t:
        i = re.search('\(',t)
        return t[i.end()::]
    else:
        return t
```

```python
df['Won'] = df['Won'].apply(winner)
```
