```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
tlkm = pd.read_csv("/kaggle/input/indonesia-popular-stocks/10_popular_indonesian_stock/TLKM.JK.csv") 
tlkm.head(5)
```

```python
tlkm.describe()
```

```python
# Select only the important features i.e. the date and price
tlkm = tlkm[["Date","Close"]] # select Date and Price
# Rename the features: These names are NEEDED for the model fitting
tlkm = tlkm.rename(columns = {"Date":"ds","Close":"y"}) #renaming the columns of the dataset
tlkm.head(5)
```

```python
from fbprophet import Prophet
mtlkm = Prophet(daily_seasonality = True) # the Prophet class (model)
mtlkm = Prophet(daily_seasonality = True) # the Prophet class (model)
mtlkm.fit(tlkm) # fit the model using all data
```

```python
type(mtlkm)
```

```python
future = mtlkm.make_future_dataframe(periods=365) #we need to specify the number of days in future
prediction = mtlkm.predict(future)
mtlkm.plot(prediction)
plt.title("Prediction of the TLKM Stock Price using the Prophet")
plt.xlabel("Date")
plt.ylabel("Close Stock Price")
plt.show()
```

```python
mtlkm.plot_components(prediction)
plt.show()
```
