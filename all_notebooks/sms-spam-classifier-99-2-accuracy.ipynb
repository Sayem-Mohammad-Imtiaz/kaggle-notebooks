# Overview
The SMS spam collection dataset is a set of SMS tagged messages, which are tagged according to being ham(legitimate) or spam. In this kernel various types models have been used to classify a message as spam or ham and the model with best accuracy (Multinomial Naive Bayes) gives a accuracy ~99.2%

```python
data=pd.read_csv('/kaggle/input/sms-spam-collection-dataset/spam.csv',encoding='latin-1')
data=data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1)
data['v1']=np.where(data['v1']=='spam',1,0)
data
```

```python
data.isnull().sum()
```

```python
plt.figure()
sns.countplot(data['v1'])
plt.show()
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data['v2'],data['v1'],random_state=0)

from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer()
X_train_cv=cv.fit_transform(X_train)
X_test_cv=cv.transform(X_test)
```

# MultinomialNB

```python

from sklearn.naive_bayes import MultinomialNB

clf_multi=MultinomialNB(alpha=0.1)
clf_multi.fit(X_train_cv,y_train)
predicted=clf_multi.predict(X_test_cv)
score=accuracy_score(y_test,predicted)
score
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression

clf=LogisticRegression()
clf.fit(X_train_cv,y_train)
predicted=clf.predict(X_test_cv)
score=accuracy_score(y_test,predicted)
score
```

# SVC

```python
from sklearn.svm import SVC

clf=SVC(kernel='linear')
clf.fit(X_train_cv,y_train)
predicted=clf.predict(X_test_cv)
score=accuracy_score(y_test,predicted)
score
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier

clf=RandomForestClassifier(random_state=0)
clf.fit(X_train_cv,y_train)
predicted=clf.predict(X_test_cv)
score=accuracy_score(y_test,predicted)
score
```

 We conclude that the best model is MultinomialNB with alpha=0.1 and an accuracy of ~99.2%
