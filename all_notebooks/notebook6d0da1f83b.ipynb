```python
yalibiaoge='/kaggle/input/arizona-houses-2021/AZhousingData.csv'
yalishuju=pd.read_csv(yalibiaoge)
print(yalishuju.columns)

yalishuju=yalishuju.dropna(axis=0)

y=yalishuju.Price
```

设置变量名yalibiaoge为表格文件 设置yalishuju变量名来读取这个文件 用columns查看这个文件的列 设置把没有属性值的行删除 设置变量y为亚历山房子的单价

```python
asdasd=['beds', 'baths', 'sqft']
X=yalishuju[asdasd]
```

设置变量asdasd是需要提取的数据 设置X变量引用这个数据

```python
from sklearn.tree import DecisionTreeRegressor
yalimoxing=DecisionTreeRegressor(random_state=250)
yalimoxing.fit(X,y)

```

从sklearn库的tree类中导入 DecisionTreeRegressor(决策树累加器)方法
定义模型。为参数 随机_状态 指定一个数字，以确保每次运行的结果相同
拟合模型

```python
print(yalimoxing.predict(X.head()))
# 229900.    294900.    683100.    260000.    290900.    377900.
```

用predict来获得预测结果

```python
from sklearn.metrics import mean_absolute_error

from sklearn.model_selection import train_test_split

# 将特性X和目标y的数据拆分为训练和验证数据
# 拆分功能基于随机数生成器。将数值提供给
# random_state参数保证我们每次使用
# 运行此脚本

# 分别设定四个变量 训练特性X,验证特性X,训练目标y,验证目标y 并将拆分结果交给它们
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 20)
# 定义模型
yalimoxing = DecisionTreeRegressor()
# 拟合模型
yalimoxing.fit(train_X, train_y)
# 根据验证数据获取预测价格结果
val_predictions = yalimoxing.predict(val_X)
# 输出绝对误差平均值，对照参数是验证目标值val_y
print(mean_absolute_error(val_y, val_predictions))
```
