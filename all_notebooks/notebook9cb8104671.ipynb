# #Visulize Data ****

# univariate pd (D1)****

```python

#Bar
plt.figure(figsize = (3,5))
bk_sh['weekday'].value_counts().sort_index().plot.bar()
plt.show()
```

```python
#line
plt.figure(figsize = (3,5))
bk_sh['mnth'].value_counts().sort_index().plot.line()
plt.show()
```

```python
#histogram
plt.figure(figsize =(3,5))
bk_sh['cnt'].hist()
plt.show()
```

# Bivariate pd (2D)****

```python
#SCATTER PLOT
plt.figure(figsize =(4,5))
bk_sh.plot.scatter(x='cnt', y='temp')
plt.show()
```

```python
#SCATTER PLOT
plt.figure(figsize =(4,3))
bk_sh.plot.scatter(x='windspeed', y='temp')
plt.show()
```

```python
#HEXPLOT
plt.figure(figsize =(3,4))
bk_sh.plot.hexbin(x='cnt', y='hum' ,gridsize = 10)
plt.show()
```

# Multivariate (3D)****

```python
#Multivariate scatter plots
sns.lmplot(x='mnth',y='atemp' ,hue = 'yr' ,data=bk_sh , fit_reg=False)
plt.show()
```

# Correlation Heatmap****

```python
plt.figure(figsize = (10,10))
sns.heatmap(bk_sh.corr() , annot=True)
plt.show()
```

```python
plt.figure(figsize = (10,10))
sns.barplot(data = bk_sh , x='holiday' , y= 'cnt' , hue='season')
plt.show()
```
