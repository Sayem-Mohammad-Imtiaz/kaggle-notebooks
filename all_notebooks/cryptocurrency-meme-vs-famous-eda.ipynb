# **Import Libraries üìö**

```python
import numpy as np 
import pandas as pd
import os
import glob
import datetime

import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True) 

sns.set(style='darkgrid')
sns.set(palette="Set2")
```

# **Load the Data üìù**

## **Meme Crypto-Currencies**

```python
for data_file in glob.glob('/kaggle/input/meme-cryptocurrency-historical-data/Meme Coin/*'):
    df_name = 'df_meme_' + os.path.basename(data_file).split('.')[0]
    globals()[df_name] = pd.read_csv(data_file)
    globals()[df_name]['Date'] = pd.to_datetime(globals()[df_name]['Date'], format='%d-%m-%Y')
```

## **Famous Crypto-Currencies**

```python
for data_file in glob.glob('/kaggle/input/meme-cryptocurrency-historical-data/*'):
    df_name = 'df_' + os.path.basename(data_file).split('.')[0]
    try:
        globals()[df_name] = pd.read_csv(data_file)
        globals()[df_name]['Date'] = pd.to_datetime(globals()[df_name]['Date'], format='%d-%m-%Y')
    except:
        continue
```

## **The columns in each file are -**

1. Date
1. Open
1. High
1. Low
1. Close
1. Volume
1. Market Cap

# **EDA üìä**

```python
def plot_timeseries_data(df, name):
    year = datetime.date.today().year
    month = datetime.date.today().month
    day = datetime.date.today().day
    initial_range = [str(year-1) + '-' + str(month) + '-' + str(day),  str(year) + '-' + str(month) + '-' + str(day+1)]
    
    if (df['Date'].min().year < 2020):
        temp = max(df['Date'].min(), datetime.datetime.strptime(initial_range[0],'%Y-%m-%d'))
        initial_range[0] = str(temp.year) + '-' + str(temp.month) + '-' + str(temp.day)

        fig1 = px.line(df.sort_values(by='Date'), x='Date', y=['High', 'Low'], )
        fig2 = px.line(df.sort_values(by='Date'), x='Date', y='Volume')

        fig1.update_layout(title = {'text' : "High-Low of "+str(name), 
                                    'y':0.95,
                                    'x':0.5,
                                    'xanchor': 'center',
                                    'yanchor': 'top',
                                    'font' : {'color' : 'Red'}})
        fig2.update_layout(title = {'text' : "Volume of "+str(name), 
                                    'y':0.95,
                                    'x':0.5,
                                    'xanchor': 'center',
                                    'yanchor': 'top',
                                    'font' : {'color' : 'Red'}})

        fig1.update_xaxes(rangeslider_visible=True)
        fig2.update_xaxes(rangeslider_visible=True)

    #     fig1.update_xaxes(rangeslider_visible=True,
    #                         rangeselector=dict(
    #                         buttons=list([
    #                             dict(count=1, label="1m", step="month", stepmode="backward"),
    #                             dict(count=6, label="6m", step="month", stepmode="backward"),
    #                             dict(count=1, label="1y", step="year", stepmode="backward"),
    #                             dict(step="all")])))
    #     fig2.update_xaxes(rangeslider_visible=True,
    #                         rangeselector=dict(
    #                         buttons=list([
    #                             dict(count=1, label="1m", step="month", stepmode="backward"),
    #                             dict(count=6, label="6m", step="month", stepmode="backward"),
    #                             dict(count=1, label="1y", step="year", stepmode="backward"),
    #                             dict(step="all")])))

        fig1['layout']['xaxis'].update(range=initial_range)
        fig2['layout']['xaxis'].update(range=initial_range)
        plotly.offline.iplot(fig1)
        plotly.offline.iplot(fig2)
    else:
        return
```

## **Famous Crypto-Currencies**

```python
for var in set(globals()):
    if (('df_' in var) and ('meme_' not in var) and (var != 'df_name')):
        plot_timeseries_data(globals()[var], var.split('.csv')[0].split('df_')[-1])
```

## **Meme Crypto-Currencies**

```python
for var in set(globals()):
    if (('df_' in var) and ('meme_' in var)):
        plot_timeseries_data(globals()[var], var.split('.csv')[0].split('df_meme_')[-1])
```
