```python
# Import library
from pycaret.classification import *
```

```python
# Load data and show the variables
df = pd.read_csv("/kaggle/input/heart-attack-analysis-prediction-dataset/heart.csv")
df.info()
```

The target variable is "output".

```python
# EDA
from pandas_profiling import ProfileReport
profile = ProfileReport(df, title="Hear Attack Data Report")
profile.to_widgets()
```

```python
# Data Preprocessing
from pycaret.classification import *
classifier_model= setup(df,
                        target='output', # Rarget variable
                        fix_imbalance=True, # Apply SMOTE
                        remove_multicollinearity=True, # Drop features with high VIF
                        ignore_low_variance=True,
                        session_id=10003, 
                        log_experiment=True, 
                        experiment_name='Heart Attack Prediction')
```

```python
# Show the model performance
best_model = compare_models()
```

### Let's see Random Forest in detail.

```python
# Create Forest Classifier
rf_clf = create_model('rf')
```

```python
# Model Tuning
tuned_rf = tune_model(rf_clf)
```

```python
# Show the parameters
tuned_rf.get_params
```

```python
# Show Confusion Matrix
plot_model(tuned_rf, plot = 'confusion_matrix')
```

```python
# Depict ROC Curve
plot_model(tuned_rf)
```

```python
# Model evaluation with various metrics
evaluate_model(tuned_rf)
```

```python
# Interpret the model 
interpret_model(tuned_rf)
```

```python
# Model interpretation with SHAP
interpret_model(tuned_rf, plot='reason', observation=15)
```

```python
# Model interpretation with correlation
interpret_model(tuned_rf, plot='correlation', observation=100)
```
