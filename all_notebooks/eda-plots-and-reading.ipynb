```python
import pandas as ps
import seaborn as sn
import matplotlib.pyplot as plt
from pandas_profiling import ProfileReport

# reading the dataset(csv file)


placements = ps.read_csv('../input/engineering-placements-prediction/collegePlace.csv')
placements.head(10)
```

```python
placements.info
```

**The "Gender" and "Stream" columns are strings( may need transformation for modelling) while PlacedOrNot column is set to binary(yaay!). Also the dataset holds 2966 rows x 8 columns**

```python
# DF describe
placements.describe()
```

```python
#The minimum and maximum values so far show signs of clarity(?). Let's see further
```

```python
pandas_profile_key = ProfileReport(placements,title="Engineering Placements Report")
pandas_profile_key.to_notebook_iframe()
```

```python
# EDA Begins...

# Comparing stream vs placed ratio plot
plt.figure(figsize=(15,5))
Stream_vs_Place = placements.groupby('Stream')['PlacedOrNot'].sum()
sn.barplot(x=Stream_vs_Place.index,y=Stream_vs_Place.values,palette = "Blues")
```

```python
# Onto comparing gender vs placed ratio plot
Gen_vs_Place = placements.groupby('Gender')['PlacedOrNot'].sum()
plt.figure(figsize=(8,5))
barplot1 = sn.barplot(x=Gen_vs_Place.index,y=Gen_vs_Place.values,palette = "crest_r")
barplot1.set(xlabel="Gender", ylabel = "Placements")
```

```python
# Onto comparing Hostel vs placed ratio plot
Hostel_vs_Place = placements.groupby('Hostel')['PlacedOrNot'].sum()
plt.figure(figsize=(8,5))
barplot2 = sn.barplot(x=Hostel_vs_Place.index,y=Hostel_vs_Place.values,palette = "Oranges")
barplot2.set(xlabel="Hostel", ylabel = "Placements")
```

```python
# Onto comparing CGPA vs placed ratio plot, Pandas Profile report showed high correlation,what's that all about ?

cgpa_vs_Place = placements.groupby('CGPA')['PlacedOrNot'].sum()
plt.figure(figsize=(8,5))
barplot3 = sn.barplot(x=cgpa_vs_Place.index,y=cgpa_vs_Place.values,palette = "Oranges_r")
barplot3.set(xlabel="CGPA", ylabel = "Placements")

```

```python
# Same relationship with a Pie plot.

plt.figure(figsize=(28,11))
pie_chart1 = cgpa_vs_Place.plot.pie(autopct="%.1f%%")
```

# The above give a 1-1 co-relation and provide awareness as we compare independent variables across competitive labels. Before proceeding to heatmaps and modelling, need a bit of label transformation for string based columns.

```python
# Quickest ways to replace labelled values is Pandas' get dummies.

placements_fe = ps.get_dummies(placements)
print(placements_fe.head(10))
```

```python
# Onto heatmap
plt.figure(figsize=(28,21))
sn.heatmap(placements_fe.corr(),cmap = "gist_heat")
```
