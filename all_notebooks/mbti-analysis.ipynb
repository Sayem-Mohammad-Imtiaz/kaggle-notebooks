## Do you know about MBTI test?, MBTI[Myers-Briggs Type Indicator] testing is a self-reported personality type test tool based on the psychological typology of Carl Jung, a Swiss psychoanalyst. It is classified into one of 16 psychological types by the results of four classification criteria. It consists of E - I indicators that show the direction of mental energy, S - N indicators that show the functions of perception including information collection, T - F indicators that make rational decisions based on information collected, and J-P indicators that show the lifestyle applied in real life.

## 1. What is the MBTI that accounts for the largest proportion?

### Let's find out what the most common MBTI is when calculating with the country's mbti ratio, not considering the number of people in each country.

```python
a = types.Type
for i in range(len(a)):
    df[f'{a[i]}'] = df[f'{a[i]}-T'] + df[f'{a[i]}-A']

country = df["Country"]
df = df[a]
df["country"]=country

b=[]
for i in range(len(a)):
    b.append(df[a[i]].sum())
    
mbti = pd.DataFrame(a)
mbti["count_per"] = b
mbti["count_per"]= round(mbti["count_per"] * 100 /158,2)

mbti = mbti.sort_values(by="count_per",ascending=False)

plt.figure(figsize=(20,7))
ax = sns.barplot(data = mbti, x="Type", y= "count_per")
for p in ax.patches:
    left, bottom, width, height = p.get_bbox().bounds
    ax.annotate("%.1f%%"%(height), (left+width/2, height*1.01), ha='center')
plt.box(False)
plt.show()
```

### Check the Barchart. 
### The most common MBTI is INFP and then ENFP, ESFJ. the INFP records 13.6%
### The smallest number of MBTI's ever records is ISTP and then ESTP. ISTP records 2.6%
