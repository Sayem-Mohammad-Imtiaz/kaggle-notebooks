```python
def height(inp):
    h_foot = float(inp.split("\'")[0])
    h_inch = float(inp.split("\'")[1])
    h_inch += (h_foot)*12
    #print(inp, h_inch, h_foot)
    h_cm = round(h_inch * 2.54, 1)
    return float(h_cm)
    
def weight(inp):
    return float(inp[:-3])
```
