```python
df.drop('EmployeeCount',axis=1,inplace=True)
df.drop('EmployeeNumber',axis=1,inplace=True)
df.drop('Over18',axis=1,inplace=True)
df.drop('StandardHours',axis=1,inplace=True)
```

```python
df.head()
```

```python
df_number_columns = df.select_dtypes(include=np.number).columns
df_number_columns
```

```python
df_category_columns = df.select_dtypes(exclude=np.number).columns
df_category_columns
```

```python
data_to_encode = df[df_category_columns]
data_to_encode
```

```python
data_encoded = pd.get_dummies(data_to_encode, drop_first = True)

data_encoded.shape
```

```python
data_encoded.info()
```

```python
df
```

```python
df[df_number_columns][:3]
```

```python
data_encoded[:5]
```

```python
final_df = pd.concat([df[df_number_columns], data_encoded], axis = 1) 
final_df.columns
```

```python
x = final_df.drop('Attrition_Yes',axis=1)
y = final_df['Attrition_Yes']
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler

min_max = MinMaxScaler()

x_scaled = min_max.fit(x).transform(x)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

x_train_scaled,x_test_scaled,y_train_scaled,y_test_scaled = train_test_split(x_scaled,y,test_size=0.25,random_state=25)
```

```python
print("training shape: ",x_train_scaled.shape)
print("testing shape: ",x_test_scaled.shape)
```

```python
# Model
model = LogisticRegression()
model.fit(x_train_scaled, y_train_scaled)
```

```python
train_pred = model.predict(x_train_scaled)
test_pred = model.predict(x_test_scaled)
```

```python
def metric_score(actual, pred):
    print('confusion matrix: ', confusion_matrix(actual, pred))
    print('accuracy score: ', accuracy_score(actual, pred))
    print('recall_score: ', recall_score(actual, pred))
    print('precision score: ', precision_score(actual, pred))
    print('f1 score: ', f1_score(actual, pred))
    print('roc auc score: ', roc_auc_score(actual, pred))
    print('classification report : ', classification_report(actual, pred))
    print('R2: ', metrics.r2_score(actual, pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, recall_score, precision_score, f1_score, roc_auc_score
from sklearn.metrics import plot_roc_curve
from sklearn import metrics
metric_score(y_train_scaled, train_pred)
```

```python
metric_score(y_test_scaled, test_pred)
```

```python
from sklearn.neighbors import KNeighborsClassifier  
KNN = KNeighborsClassifier(n_neighbors = 5, metric = 'manhattan', weights = 'uniform', n_jobs = -1)
KNN.fit(x_train_scaled, y_train_scaled)
predict_test = KNN.predict(x_test_scaled)
predict_train = KNN.predict(x_train_scaled)
predict_train
```

```python
metric_score(y_train_scaled, predict_train)
```

```python
metric_score(y_test_scaled, predict_test)
```
