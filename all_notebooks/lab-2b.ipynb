```python
data = pd.read_csv('/kaggle/input/oranges-vs-grapefruit/citrus.csv')
data.head(10)
```

```python
data.describe()
```

```python
sns.heatmap(data.corr(),cmap='coolwarm')
```

```python
%matplotlib inline
sns.pairplot(data,hue='name')
```

```python
y=data.name
X=data.drop(['name'],axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

```

```python
#ф активации макс(0,х), много нейронов в 3 слоя
clf = MLPClassifier(hidden_layer_sizes=(1000,100,20),activation='relu',solver='adam',learning_rate='adaptive', learning_rate_init=0.001,max_iter=500,shuffle=True,random_state=1)
clf.fit(X_train, y_train)
predic=clf.predict(X_test)
clf.score(X_test,y_test)
print(accuracy_score(y_test,predic))
confusion_matrix(y_test,predic)
```

```python
print(classification_report(y_test,predic))
```

```python
#ф активации сигмоида
clf = MLPClassifier(hidden_layer_sizes=(20),activation='logistic',solver='sgd',learning_rate='adaptive', learning_rate_init=0.001,max_iter=500,shuffle=True,random_state=0)
clf.fit(X_train, y_train)
predic=clf.predict(X_test)
clf.score(X_test,y_test)
print(accuracy_score(y_test,predic))
confusion_matrix(y_test,predic)
print(classification_report(y_test,predic))
```

```python
#ф активации гип_тангенс, 6 нейронов
clf = MLPClassifier(hidden_layer_sizes=(6),activation='tanh',solver='adam',learning_rate='adaptive', learning_rate_init=0.001,max_iter=300,shuffle=True,random_state=0)
clf.fit(X_train, y_train)
predic=clf.predict(X_test)
clf.score(X_test,y_test)
print(accuracy_score(y_test,predic))
confusion_matrix(y_test,predic)
print(classification_report(y_test,predic))
```

```python
#ф активации max(0,x), 1 нейрон
clf = MLPClassifier(hidden_layer_sizes=(5),activation='relu',solver='adam',learning_rate='constant', learning_rate_init=0.01,max_iter=200,shuffle=True,random_state=1)
clf.fit(X_train, y_train)
predic=clf.predict(X_test)
clf.score(X_test,y_test)
print(accuracy_score(y_test,predic))
confusion_matrix(y_test,predic)
print(classification_report(y_test,predic))
```
