# Logistic Regression

```python
# Putting feature variable to X
X = dib['Blood Sugar Level']

# Putting response variable to y
y = dib['Diabetes']
```

```python
import statsmodels.api as sm
```

```python
logm1 = sm.GLM(y,(sm.add_constant(X)), family = sm.families.Binomial())
result=logm1.fit().summary()
result
```
