```python
trainp = "/kaggle/input/heits-weits/heitwt.csv"
navals = ["?","",None,"None",np.nan]
df = pd.read_csv(trainp,na_values=navals)
df.shape
```

```python
X=df.values
```

```python
iner=list()
for i in range(1,13):
    model = KMeans(n_clusters=i,random_state=0)
    model.fit(X)
    iner.append(model.inertia_)
```

```python
plt.scatter(range(1,13),iner)
plt.plot(range(1,13),iner)
```

```python
model=KMeans(n_clusters=2,random_state=0)
model.fit(X)
y=model.predict(X)
y
```

```python
model.cluster_centers_
```
