# 0. Data Cleaning

```python
df=df.drop('Unnamed: 0', axis=1)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(how='any')
```

```python
df.dtypes
```

```python
from datetime import datetime as dt
```

```python
df['Employment_Start_Date']=pd.to_datetime(df['Employment_Start_Date'])
df['Employment_End_Date']=pd.to_datetime(df['Employment_End_Date'])
df['Employment_Start_Date']=df['Employment_Start_Date'].dt.strftime('%Y-%m-%d')
df['Employment_End_Date']=df['Employment_End_Date'].dt.strftime('%Y-%m-%d')
```

```python
df['Employment_period']=df['Employment_period'].str.replace(' days', '')
df['Employment_period']=df['Employment_period'].astype(int)
```

```python
df=df[df.Employment_period > 0]
```

```python
df.describe()
```

# 2. Employment Duration

# 2-1.  Years

```python
from decimal import Decimal, ROUND_HALF_UP, ROUND_HALF_EVEN
```

```python
df['period_y'] = df['Employment_period']/365
```

```python
def period(ex):
    if ex<1:
        return 0
    elif 1<=ex<2:
        return 1
    elif 2<=ex<3:
        return 2
    elif 3<=ex<4:
        return 3
    elif 4<=ex<5:
        return 4
    else:
        return 5
```

```python
df['period']=df['period_y'].apply(period)
```

```python
plt.figure(figsize=(10, 10))
sns.countplot(y='period',data=df)
```

# 3. Engineering Degrees
