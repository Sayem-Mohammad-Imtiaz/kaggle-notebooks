importing 2 datasets

```python
banglore = pd.read_csv('../input/housing-prices-in-metropolitan-areas-of-india/Bangalore.csv')
mumbai = pd.read_csv('../input/housing-prices-in-metropolitan-areas-of-india/Mumbai.csv')
```

```python
banglore.head(5)
```

```python
mumbai.head(5)
mumbai.columns
```

Since 0 and 1 tells us that the house comes with amenities or not and 9 in the data means not mentioned. So we will convert all 9 to nan and then drop all the missing values.

```python
banglore.replace(9,np.nan,inplace=True)
mumbai.replace(9,np.nan,inplace=True)
```

```python
banglore = banglore.dropna()
mumbai = mumbai.dropna()
```

Changing Price to Price in lakhs

```python
banglore['Price'] = banglore['Price']/100000
mumbai['Price'] = mumbai['Price']/100000
```

```python
banglore.Price.head()
```

We will start with data visualization.

```python
plt.figure(figsize=(12,6))
plt.title('Area in sq.feet vs Price (in lakh)')
sns.scatterplot(x=banglore['Price'],y=banglore['Area'],hue=banglore['No. of Bedrooms'])
```

```python
plt.figure(figsize=(12,6))
plt.title('Area in sq.feet vs Price (in lakh)')
sns.scatterplot(x=mumbai['Price'],y=mumbai['Area'],hue=mumbai['No. of Bedrooms'])
```

```python
#df = mumbai.Location.value_counts().index[:10]
plt.figure(figsize=(12,10))
plt.title('Total Properties in Mumbai (According to maximum available properties)')
sns.countplot(y='Location',data=mumbai,order= mumbai.Location.value_counts().index[:10],palette = "Set3")
```

```python
plt.figure(figsize=(12,10))
plt.title('Total Properties in Banglore (According to maximum available properties)')
sns.countplot(y='Location',data=banglore,order= banglore.Location.value_counts().index[:10],palette = "Set3")
```

```python
banglore_mod = banglore[['Price','Area','No. of Bedrooms','Gymnasium','SwimmingPool','IndoorGames','JoggingTrack']]
mumbai_mod = mumbai[['Price','Area','No. of Bedrooms','Gymnasium','SwimmingPool','IndoorGames','JoggingTrack']]
```

```python
frames = [banglore_mod,mumbai_mod]
merged = pd.concat(frames)
plt.figure(figsize=(15,8))
sns.heatmap(merged.corr(),annot=True)
```
