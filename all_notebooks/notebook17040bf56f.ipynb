```python
d_new=d[d['Fuel Consumption City (L/100 km)']<d['Fuel Consumption City (L/100 km)'].quantile(0.912)]
d_new=d_new[d_new['CO2 Emissions(g/km)']<d_new['CO2 Emissions(g/km)'].quantile(0.99)]
print(d_new.shape)
d.shape
```

```python
#actual data
Y=d['CO2 Emissions(g/km)'].values.reshape(-1,1)
X=d['Fuel Consumption City (L/100 km)'].values.reshape(-1,1)
#data after removing outliers
y=d_new['CO2 Emissions(g/km)'].values.reshape(-1,1)
x=d_new['Fuel Consumption City (L/100 km)'].values.reshape(-1,1)
X.min(),X.max(),Y.min(),Y.max()
```

```python
x.min(),x.max(),y.min(),y.max()
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X,Y)
lr.fit(x,y)
```

```python
Y_pred=lr.predict(X)
Y_pred
```

```python
y_pred=lr.predict(x)
y_pred
```

```python
#cost_function(with outliers)
from sklearn.metrics import mean_squared_error, mean_absolute_error
mean_absolute_error(Y,Y_pred),np.sqrt(mean_squared_error(Y,Y_pred))
```

```python
#with outliers
plt.scatter(X,Y)
plt.scatter(X,Y_pred)
```

```python
#cost_function(without outliers)
from sklearn.metrics import mean_squared_error, mean_absolute_error
mean_absolute_error(y,y_pred),np.sqrt(mean_squared_error(y,y_pred))
```

```python
#data after removing outliers
plt.scatter(x,y)
plt.scatter(x,y_pred)
```

```python
#training set & test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
lr.fit(X_train,Y_train)
Y_train_pred=lr.predict(X_train)
print(Y_train)
Y_train_pred
```

```python
#train set
lr.fit(x_train,y_train)
y_train_pred=lr.predict(x_train)
print(y_train)
y_train_pred
```

```python
mean_absolute_error(Y_train,Y_train_pred)
```

```python
mean_absolute_error(y_train,y_train_pred)
```

```python
#train set accuracy
from sklearn.metrics import r2_score
r2_train_accX = r2_score(Y_train, Y_train_pred)
print("Accuracy:", r2_train_accX*100)
```

```python
r2_train_accx = r2_score(y_train, y_train_pred)
print("Accuracy:", r2_train_accx*100)
```
