```python
import nltk
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer,HashingVectorizer,TfidfTransformer
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix,accuracy_score
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
import matplotlib.pyplot as plt
```

```python
df.head()
```

```python
df['content'] = df['content'].astype(str)
```

```python

def unicode_to_ascii(s):
    return ''.join(c for c in unicodedata.normalize('NFD', s) if unicodedata.category(c) != 'Mn')

def clean_stopwords_shortwords(w):
    stopwords_list=stopwords.words('english')
    words = w.split() 
    clean_words = [word for word in words if (word not in stopwords_list) and len(word) > 2]
    return " ".join(clean_words) 

def preprocess_sentence(w):
    w = unicode_to_ascii(w.lower().strip())
    w = re.sub(r"([?.!,多])", r" ", w)
    w = re.sub(r'[" "]+', " ", w)
    w = re.sub(r"[^a-zA-Z?.!,多]+", " ", w)
    w=clean_stopwords_shortwords(w)
    w=re.sub(r'@\w+', '',w)
    return w
```

```python
X = df['content']
#y = pd.get_dummies(df['score']).values
messages = X.copy()
messages = list(messages)
```

```python
import unicodedata
```

```python
voc_size = len(X)

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = preprocess_sentence(messages[i])
    corpus.append(review)
```

```python
# CountVectorizer
cv = CountVectorizer(max_features=4000,ngram_range=(1,4))
X = cv.fit_transform(corpus).toarray()
cv.get_feature_names()[:20]
```

```python
import pickle
pickle.dump(cv, open("CountVectorizer.pickel", "wb"))

```

```python
cv.get_params()
```

```python
y = df['score']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,stratify = df['score'], random_state=42)
```

# # Multinomial NaiveBayes

```python
# Multinomial NaiveBayes
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(X_train,y_train)
y_pred = mnb.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print("naive bayes accuracy is ",round(accuracy_score(y_test,y_pred),2)*100,"%")
print("\n")
print("confusion matrix for naive bayes ")
plot_confusion_matrix(cm,class_names=['extremely negative','negative','neutral','positive','extremely positive'],cmap='Purples',show_normed=True,colorbar=True,figsize=(6,6))
plt.show()
```

```python
import pickle
Pkl_Filename = "count_MNB_FINAL.pkl"  

with open(Pkl_Filename, 'wb') as file:  
    pickle.dump(mnb, file)
```

# PassiveAgressiveClassifier

```python
# PassiveAgressiveClassifier
from sklearn.linear_model import PassiveAggressiveClassifier
pac = PassiveAggressiveClassifier()
pac.fit(X_train,y_train)
y_pred = pac.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print("Passive Aggressive Classifier accuracy is ",round(accuracy_score(y_test,y_pred),2)*100,"%")
print("\n")
print("confusion matrix for Passive Aggressive Classifier ")
plot_confusion_matrix(cm,class_names=['extremely negative','negative','neutral','positive','extremely positive'],cmap='Purples',show_normed=True,colorbar=True,figsize=(6,6))
plt.savefig("PASAGG_count.png")
plt.show()
```

```python
import nltk
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer,HashingVectorizer,TfidfTransformer
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix,accuracy_score
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
import matplotlib.pyplot as plt
```

```python
df['content'] = df['content'].astype(str)
```

```python
def unicode_to_ascii(s):
    return ''.join(c for c in unicodedata.normalize('NFD', s) if unicodedata.category(c) != 'Mn')

def clean_stopwords_shortwords(w):
    stopwords_list=stopwords.words('english')
    words = w.split() 
    clean_words = [word for word in words if (word not in stopwords_list) and len(word) > 2]
    return " ".join(clean_words) 

def preprocess_sentence(w):
    w = unicode_to_ascii(w.lower().strip())
    w = re.sub(r"([?.!,多])", r" ", w)
    w = re.sub(r'[" "]+', " ", w)
    w = re.sub(r"[^a-zA-Z?.!,多]+", " ", w)
    w=clean_stopwords_shortwords(w)
    w=re.sub(r'@\w+', '',w)
    return w
```

```python
X = df['content']
#y = pd.get_dummies(df['score']).values
messages = X.copy()
messages = list(messages)
```

```python
import unicodedata
voc_size = len(X)

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = preprocess_sentence(messages[i])
    corpus.append(review)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
```

```python
tfv = TfidfVectorizer(min_df=3,  max_features=4000, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 4), use_idf=1,smooth_idf=1,sublinear_tf=1,
            stop_words = 'english')

X = tfv.fit_transform(corpus).toarray()
tfv.get_feature_names()[:20]
tfv.get_params()
y = df['score']
```

```python
import pickle
pickle.dump(tfv, open("TFIDF.pickel", "wb"))

```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,stratify = df['score'],random_state=42)
```

# MNB

```python
# Multinomial NaiveBayes
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
mnb.fit(X_train,y_train)
y_pred = mnb.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print("naive bayes accuracy is ",round(accuracy_score(y_test,y_pred),2)*100,"%")
print("\n")
print("confusion matrix for naive bayes ")
plot_confusion_matrix(cm,class_names=['extremely negative','negative','neutral','positive','extremely positive'],cmap='Purples',show_normed=True,colorbar=True,figsize=(6,6))
plt.savefig("MNB_TFIDF_FINAL")
plt.show()
```

```python
import pickle
Pkl_Filename = "TFIDF_MNB_FINAL.pkl"  

with open(Pkl_Filename, 'wb') as file:  
    pickle.dump(mnb, file)
```

# PASSIVE AGGRESSIVE

```python
# PassiveAgressiveClassifier
from sklearn.linear_model import PassiveAggressiveClassifier
pac = PassiveAggressiveClassifier()
pac.fit(X_train,y_train)
y_pred = pac.predict(X_test)
cm = confusion_matrix(y_test,y_pred)
print("Passive Aggressive Classifier accuracy is ",round(accuracy_score(y_test,y_pred),2)*100,"%")
print("\n")
print("confusion matrix for Passive Aggressive Classifier ")
plot_confusion_matrix(cm,class_names=['extremely negative','negative','neutral','positive','extremely positive'],cmap='Purples',show_normed=True,colorbar=True,figsize=(6,6))
plt.savefig("TFIDF_PASAG_FINAL.png")
plt.show()
```
