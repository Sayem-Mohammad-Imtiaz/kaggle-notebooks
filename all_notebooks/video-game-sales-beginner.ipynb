# **Video Game Sales Analysis**

Importing the essential modules,

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading the dataset:

```python
data='../input/videogamesales/vgsales.csv'
vg_sales= pd.read_csv(data, header=0, sep=',')
print(vg_sales.head())
```

```python
print(vg_sales.describe())
```

```python
print(vg_sales.info())
```

Finding the total null values in each column:

```python
print(vg_sales.isnull().sum())
```

For analysing sales with respect to the year, we must fill the null cells.

```python
mean = vg_sales['Year'].mean()
vg_sales['Year'].fillna(mean, inplace=True)
plt.xticks(rotation = 75)
x_axis = vg_sales['Year'].astype(int)
sns.countplot(x= x_axis, data = vg_sales)
plt.title('Total Game Sales Each Year')
plt.show()
```

**Thus, most video games were sold in *2009 and 2008***

```python
plt.xticks(rotation = 75)
x_axis = vg_sales['Genre']
sns.countplot(x= x_axis, data = vg_sales)
plt.title('Total Game Sales of Each Genre')
plt.show()
```

**Thus, *Action* is the most sold genre**

For, the most sold games in each genre:

```python
game_sales = vg_sales[['Genre','Name']].groupby(['Genre']).agg(lambda x:x.value_counts().index[0])
genre_games = game_sales.rename(columns = {'Name' : 'Game'}, inplace = False)
genre_games
```

For the most video game sales on platforms:

```python
plt.xticks(rotation = 75)
x_axis = vg_sales['Platform']
sns.countplot(x= x_axis, data = vg_sales)
plt.title('Total Game Sales on Each Platform')
plt.show()
```

**Thus, the *PS2 and the DS* are the most used platforms**

To find the correlation between the numerical values in the dataset:

```python
correlation_vg_sales = vg_sales.corr()

axis_corr = sns.heatmap(
correlation_vg_sales,
vmin=-1, vmax=1, center=0,
cmap=sns.diverging_palette(50, 500, n=500),
square=True
)

plt.show()
```

**Thus, both *NA sales and EU* sales are closely related to *Global sales***

For analysing the most common publisher, we must fill the null cells with a common value 'Unknown'

```python
vg_sales['Publisher'] = vg_sales['Publisher'].fillna('Unknown')
number_df = vg_sales.groupby('Publisher')[['Name']].count().sort_values('Name', ascending = False).head(50)
number_clean = number_df.rename(columns = {'Name' : 'Number'}, inplace = False)
number_clean
```

```python
number_df.plot(kind = 'bar', figsize = (25, 10));
plt.xlabel('Publisher', fontsize = 20);
plt.ylabel('Number of video games released', fontsize = 20);
plt.title('Top Publishers of Games', fontsize = 40)
```

**Thus, most common Pulisher of video games is *Electronic arts followed by Activision***

For most common games by region,

```python
top_games_NA = vg_sales.sort_values('NA_Sales',ascending = False).head(5)
explode = [0.1, 0, 0, 0, 0]
plt.pie(top_games_NA['NA_Sales'], labels = top_games_NA['Name'], explode = explode)
plt.show()
```

```python
top_games_EU = vg_sales.sort_values('EU_Sales',ascending = False).head(5)
explode = [0.1, 0, 0, 0, 0]
plt.pie(top_games_EU['EU_Sales'], labels = top_games_EU['Name'], explode = explode)
plt.show()
```

```python
top_games_JP = vg_sales.sort_values('JP_Sales',ascending = False).head(5)
explode = [0.1, 0, 0, 0, 0]
plt.pie(top_games_JP['JP_Sales'], labels = top_games_JP['Name'], explode = explode)
plt.show()
```

```python
top_games_Other = vg_sales.sort_values('Other_Sales',ascending = False).head(5)
explode = [0.1, 0, 0, 0, 0]
plt.pie(top_games_Other['Other_Sales'], labels = top_games_Other['Name'], explode = explode)
plt.show()
```

```python
top_games_Global = vg_sales.sort_values('Global_Sales',ascending = False).head(5)
explode = [0.1, 0, 0, 0, 0]
plt.pie(top_games_Global['Global_Sales'], labels = top_games_Global['Name'], explode = explode)
plt.show()
```
