```python
datadir = '/kaggle/input/evs-one-electric-vehicle-dataset/ElectricCarData_Clean.csv'
rawdata = pd.read_csv(datadir)
((rawdata.sample(10).set_index('Brand')).T)
```

```python
rawdata.info()
```

```python
for attribute in rawdata.columns:
    if rawdata[attribute].dtype == 'object':
        print(('Categorial Variables: {}').format(attribute), rawdata[attribute].unique(),'\n', sep='\n')
```

```python
# Univariate EDA - historgram
for attribute in rawdata.columns:
    if rawdata[attribute].dtype != 'object':
        fig = px.histogram(rawdata[attribute], title=('{}').format(attribute))
        plotol(fig)
```
