```python
#import necessary libraries
import numpy as np
import pandas as pd
```

```python
tips_df = pd.read_csv('../input/tips-dataset/tips.csv')
tips_df.head()
```

```python
tips_df.info()
```

there is no missing data in this data frame

```python
# size shows number of people for each total_bill 
# we can find the average bill that has been paid for each person by dividing the total bill by the size
tips_df['pay_per_person'] = np.round((tips_df['Total_bill'] / tips_df['Size']), 2)
tips_df.head()
```

```python
# find the percentage of tips by total bill
tips_df['tips_percentage'] = np.round((tips_df['Tip']*100) / tips_df['Total_bill'] , 2)
tips_df.head()
```

```python
# find out how many customers paid a total bill more than minimum and they were less than 3 person
df_1 = tips_df[(tips_df['Total_bill'] > tips_df['Total_bill'].mean())  & (tips_df['Size'] <= 2)]
df_1.head()
```

```python
df_1.info()
```

so there are 39 customers with the size less than 3 and total bill more than average

```python
men_average_tip = tips_df.groupby('Sex').mean().loc['Male' , 'Tip']
women_average_tip = tips_df.groupby('Sex').mean().loc['Female' , 'Tip']
if men_average_tip > women_average_tip:
    print('men pay tips more than women')
else:
    print('women pay tips more than men')
```
