### dividing the data into train and test data set

```python
#function to divide the data into test and train data set
def split_train_test(data,ratio):
    np.random.seed(32) #start the randomness from 32
    suffled_idx = np.random.permutation(len(data))
    test_set_size = int(len(data)*ratio)
    test_idx = suffled_idx[:test_set_size]
    train_idx = suffled_idx[test_set_size:]
    
    return data.iloc[train_idx],data.iloc[test_idx]
```

split the data

```python
train_set, test_set = split_train_test(mnist,.02)

print('train',len(train_set))
print('test',len(test_set))
```

## Precision

### precision =~ accuracy

$$precision = \frac{TP}{TP + FP}$$
   

```python
from sklearn.metrics import precision_score

precision_score(Y_train_2,Y_train_pred)
```

## Recall

$$Recall = \frac{TP}{TP + FN}$$

```python
from sklearn.metrics import recall_score
recall_score(Y_train_2,Y_train_pred)
```

** imp : your classifier should have high precision and high recall **

## F1 Score

** Harmonic mean of precision and recall **
$$F1 = 2X\frac{prescision  X  recall}{precision + recall}$$

$$F1 = \frac{TP}{TP + \frac{FN + FP}{2}}$$

```python
from sklearn.metrics import f1_score
f1_score(Y_train_2,Y_train_pred)
```
