# More Popular Artists

```python
plt.figure(figsize=(9, 4))
sns.set(style="whitegrid")
x = df.groupby("artists")["popularity"].sum().sort_values(ascending=False).head(20)
ax = sns.barplot(x.index, x)
ax.set_title('Top Artists with Popularity by Sum')
ax.set_ylabel('Popularity')
ax.set_xlabel('Artists')
plt.xticks(rotation = 90)
```

# Songs by Genre

```python
df2 = pd.read_csv('/kaggle/input/spotify-dataset-19212020-160k-tracks/data_w_genres.csv')
```

```python
df2.isna().sum()
genre=df2[["genres","artists"]]
genre=genre[genre["genres"]!="[]"]
genre["genres"]=genre["genres"].str.replace("'", "")
genre["genres"]=genre["genres"].str.replace("[", "")
genre["genres"]=genre["genres"].str.replace("]", "")
```

```python
genre_dict={}

genre_df_dict=genre["genres"].str.split(",")
```

```python
genre_df_dict[0]
```

```python
for index, genre_list in genre_df_dict.iteritems():
    for element in genre_list:
        if element in genre_dict:
            genre_dict[element] +=1
        else:
            genre_dict[element] = 1
            
```

```python
genre_dict_sorted=sorted(genre_dict.items(),key=lambda x: x[1],reverse=True) 
```

```python
genre_dict_sorted[1:20]
```

```python
fig = plt.figure(figsize = (5, 5))
ax = fig.subplots()
genre.value_counts()[:10].plot(ax=ax, kind = "pie")
ax.set_ylabel("")
ax.set_title("Top 10 most popular genres")
plt.show()
```
