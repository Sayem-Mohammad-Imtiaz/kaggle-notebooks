What this project does is look at the ability of a logistic regression algorithm to predict the likelihood of a cell being cancerous based on its cytology. 

```python
data = pd.read_csv("/kaggle/input/wisconsin-breast-cancer-cytology-features/wisconsin_breast_cancer.csv")
```

Read in the dataset

```python
data.head()
```

A look at the first five rows of the dataset.

```python
data=data.dropna(how='any')
data.shape
```

Drop any rows with missing values and display the shape of the matrix.

```python
x = data[['thickness','size','shape','adhesion','single','nuclei','chromatin','nucleoli','mitosis']]
x.head()
```

Set the predictive dataframe.

```python
y = data['class']
y.head()
```

Set the target series and variable.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

Set up a cross-validation sequence to eliminate bias or overfitting of the data.

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(x_train,y_train)
```

Train the logistic regression model.

```python
y_pred_class=logreg.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred_class)
```

Now we can see that a logistic regression model based on all of the cell's cytology results in an accuracy of around 95% when predicting whether or not a cell is cancerous.

Let's see how it fairs when based off of the mitotic state of the cell.

```python
x = data[['mitosis']]
y = data['class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0)
```

```python
logreg.fit(x_train,y_train)
```

```python
y_pred_class=logreg.predict(x_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred_class)
```

So, after using only the mitotic state of the cell as the independent variable for the logistic regression algorithm, it results in an 80% accuracy rate. This indicates that the mitotic state or activity of a cell is the main determinant in whether or not a cell is cancerous.
