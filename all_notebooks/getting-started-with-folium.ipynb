```python
# Map method of folium return Map object 
  
# Here we pass coordinates of Gfg and starting Zoom level = 4

usa_map = folium.Map([39.358, -98.118], zoom_start=4)
usa_map
```

```python
#Let us print the chipotel located Forest Hills, Newyork.

forest_Hills = folium.Map([40.7204886,-73.8459915], zoom_start = 25)
map_title = 'Forest Hills, NewYork'
folium.Marker([40.7204886,-73.8459915], 
               popup = 'Forest Hills').add_to(forest_Hills) 
forest_Hills
```

```python
usa_map1 = folium.Map([39.358, -98.118], zoom_start=4)
for lat, lon,loc in zip(df.latitude, df.longitude,df.location):
    folium.Marker([lat, lon],radius=10, color=None,fill_opacity=0.3,
                        tooltip="Location : "+str(loc)).add_to(usa_map1)
usa_map1
```

```python
usa_map2 = folium.Map([39.358, -98.118], zoom_start=4)
for lat, lon,loc in zip(df.latitude, df.longitude,df.location):
    folium.CircleMarker([lat, lon],radius=10,fill_opacity=0.2,
                        tooltip="Location : "+str(loc)).add_to(usa_map2)
usa_map2
```
