```python
from sklearn.preprocessing import LabelEncoder

categorical = list(fires.select_dtypes(include = ["object"]).columns)
for i, column in enumerate(categorical) :
    label = LabelEncoder()
    fires[column] = label.fit_transform(fires[column])
```
