```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml import Pipeline
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.feature import StringIndexer
```

```python
spark = SparkSession.builder.master("local")\
        .appName("Colab_Excersie")\
        .config('spark.ui.port', '4050')\
        .getOrCreate()

print(spark)
```

```python
df=spark.read.csv('/kaggle/input/iris-dataset/Iris.csv',header=True)
```

```python
df.select("species").distinct().show()
```

```python
mod=df.select(df.SepalLengthCm.cast("float"),
              df.SepalWidthCm.cast("float"),
              df.PetalLengthCm.cast("float"),
              df.PetalWidthCm.cast("float"),df.Species)
```

```python
mod.show()
```

```python
labelEncoding = StringIndexer(inputCol= 'Species', outputCol= 'y')
vectorAssemble= VectorAssembler(inputCols=['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm'], outputCol='X')
dt = DecisionTreeClassifier(featuresCol = 'X', labelCol = 'y', maxDepth = 4)
classification_pipeline = Pipeline(stages=[labelEncoding ,vectorAssemble,dt])
model = classification_pipeline.fit(mod)
```

```python
transf_mod=model.transform(mod)
```
