```python
from sklearn import preprocessing
import numpy as np
x_array = veri.drop(["Potability"], axis=1)
normalized_arr = preprocessing.normalize(x_array)
normalized_arr.shape
```

```python
from sklearn import preprocessing
dff = veri.drop("Potability",axis=1)
x = dff.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
dff = pd.DataFrame(x_scaled)
```

# LightGBM

```python
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import mean_squared_error,r2_score
from warnings import filterwarnings
filterwarnings('ignore')
x = normalized_arr
y = veri["Potability"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
lgbm_model = LGBMClassifier().fit(x_train, y_train)
y_pred = lgbm_model.predict(x_test)
cross_val_score(lgbm_model,x_test,y_test,cv=10).mean()
```

# Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import mean_squared_error,r2_score,accuracy_score
from warnings import filterwarnings
filterwarnings('ignore')
X = dff
y = veri["Potability"]
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=123)
nb = GaussianNB()
nb_model = nb.fit(x_train, y_train)
nb_model.predict_proba(x_test)
y_pred = nb_model.predict(x_test)

accuracy_score(y_test, y_pred)
cross_val_score(nb_model, x_test, y_test, cv = 100).mean()
```
