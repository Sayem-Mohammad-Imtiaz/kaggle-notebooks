```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import re
df = pd.read_csv('../input/2010-report-to-congress-on-white-house-staff/2010_Report_to_Congress_on_White_House_Staff.csv')
sns.set(rc={'figure.figsize':(10,8)})
```

```python
total_no_of_staff = df.shape[0]
staff_status_type = df['Employee Status'].unique()
unique_positions = df['Position Title'].unique()
max_sal = df['Salary'].max()
min_sal = df['Salary'].min()
total_staff_expenditure = df['Salary'].sum()
pay_basis = df['Employee Status'].unique()
unique_employee_status  = df['Employee Status'].value_counts()

print('**********************************Summary*****************************************')
print('Total Number of Staff working in White House:', total_no_of_staff)
print('Employe status types for Staff working in White House:', ", ".join(staff_status_type))
print(unique_employee_status)
print('Total Number of unique positions:', len(unique_positions))

print('White House staff expenditure: ', total_staff_expenditure)
print('White House staff Pay Basis types: ', ", ".join(pay_basis))
print('White House staff Maximum Salary: ', max_sal)
print('White House staff Minimum Salary: ', min_sal)

print('**********************************************************************************')
```

```python
print(df.describe().round(2))
```

```python
ax = sns.countplot(x="Employee Status", data=df, linewidth=0)
plt.xlabel('Employee status')
plt.ylabel('Employee count')
plt.title('Staff Distribution By Employee status', fontdict = {'fontsize' : 25, 'color': 'blue'})


for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total_no_of_staff)
        x = p.get_x() + p.get_width() + 0.02
        y = p.get_y() + p.get_height()/2
        label = f'{p.get_height()}({percentage})'
        ax.annotate(label, (x, y))

plt.show()
```

```python
totalCount = df['Salary'].count()
# count is 469
noSalary = ((df[(df['Salary'] == 0) ]['Salary'].count())/totalCount)*100
#no salary count is 3
# count is 106
upto45k = ((df[(df['Salary'] > 0) & (df['Salary'] <= 45000)]['Salary'].count())/totalCount)*100
# count is 191
from45kto90k = ((df[(df['Salary'] > 45000) & (df['Salary'] <= 90000)]['Salary'].count())/totalCount)*100
# count is 105
from90kto135k = ((df[(df['Salary'] > 90000) & (df['Salary'] <= 135000)]['Salary'].count())/totalCount)*100
# count is 38
from135kto170k = ((df[(df['Salary'] > 135000) & (df['Salary'] <= 170000)]['Salary'].count())/totalCount)*100
# count is 26
gt170k = ((df[(df['Salary'] > 170000)]['Salary'].count())/totalCount)*100
labels = 'No salary', 'up to 45k', '45k to 90k', '90k to 135k', '135 to 170k', 'more then 170k'
sizes = [noSalary, upto45k, from45kto90k, from90kto135k, from135kto170k, gt170k]
explode = (0.2, 0, 0, 0,0,0)  # only "explode" the 2nd slice (i.e. 'Hogs')
fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=0, radius = 1.2, textprops={"fontsize":20})
#ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Salary Distribution', fontdict = {'fontsize' : 25, 'color': 'blue'})
plt.show()
```

```python
sns.set(rc={'figure.figsize':(10,8)})
sns.boxplot(y=df.Salary,x=df['Employee Status'],orient='v')
sns.violinplot(y=df.Salary,x=df['Employee Status'],orient='v')
plt.title('Box and Violin plot for Salary Distribution', fontdict = {'fontsize' : 25, 'color': 'blue'})
```
