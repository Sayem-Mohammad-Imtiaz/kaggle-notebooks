```python
df['home'] = df['result_full'].str.split('-').str[0]
```

```python
df['Away'] = df['result_full'].str.split('-').str[1]
```

```python
df.head()
```

```python
df['Winner'] = '0'
```

```python
df.head()
```

```python
df['home'] = df['home'].astype(int)
df['Away'] = df['Away'].astype(int)
```

```python
for i in range(len(df['result_full'])):
    if df['home'][i] > df['Away'][i]:
        df['Winner'][i] = 'Home'
    elif df['home'][i] == df['Away'][i]:
        df['Winner'][i] == 'Draw'
    elif df['home'][i] < df['Away'][i]:
        df['Winner'][i] = 'Away'
    else:
        df['Winner'][i] = 'No Result'
    
```

```python
df.head()
```

```python
df['Winner'] = df['Winner'].replace('0','Draw')
```

```python
df['Winner'].value_counts()
```

# MANCHESTER CITY PERFORMANCE IN LAST 10 YEARS

```python
home_goals = 0
away_goals = 0
for i in range(len(df['result_full'])):
    if df['home_team'][i] == 'Manchester City':
        home_goals = home_goals + df['home'][i]
    elif df['away_team'][i] == 'Manchester City':
        away_goals = away_goals + df['Away'][i]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14,5))
plt.title('Manchester City Goals Throughout Season')
sns.lineplot(df[df['home_team']=='Manchester City']['season'],df[df['home_team']=='Manchester City']['home'],color='blue')
sns.lineplot(df[df['away_team']=='Manchester City']['season'],df[df['away_team']=='Manchester City']['Away'],color='green')
```
