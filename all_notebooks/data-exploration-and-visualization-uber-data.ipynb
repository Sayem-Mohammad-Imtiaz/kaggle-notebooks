### Data Loading and Preparation

Loadin Data

```python
os.listdir(r'../input/uber-pickups-in-new-york-city')
```

```python
files = ['uber-raw-data-apr14.csv',
         'uber-raw-data-aug14.csv',
         'uber-raw-data-jul14.csv',
         'uber-raw-data-jun14.csv',
         'uber-raw-data-may14.csv',
         'uber-raw-data-sep14.csv']
files
```

```python
df_names = [re.findall('\w+\d',f)[0][:-2].upper() for f in files]
df_names
```

```python
#Concate all data in one dataframe

path=r'../input/uber-pickups-in-new-york-city'

apr14=pd.read_csv(path+"/"+files[0],encoding='utf-8')
aug14=pd.read_csv(path+"/"+files[1],encoding='utf-8')
jul14=pd.read_csv(path+"/"+files[2],encoding='utf-8')
jun14=pd.read_csv(path+"/"+files[3],encoding='utf-8')
may14=pd.read_csv(path+"/"+files[4],encoding='utf-8')
sep14=pd.read_csv(path+"/"+files[5],encoding='utf-8')

#blank dataframe
uber_2014=pd.DataFrame()
dfs = [apr14,aug14,jul14,jun14,may14,sep14]

for df in dfs:
    uber_2014=pd.concat([df,uber_2014])

dfs.append(uber_2014)

uber_2014.head()
```

<h2> Data Preparation</h2>
<b><p>Now we have :</b><br>
<li>Lat : The latitude of the Uber pickup
<li>Lon : The longtude of the Uber pickup
<li>Base : The TLC base company code affiliated with the Uber pickup
<li>Date/Time : The data of the trip

```python
#Preparing the columns in each Dataframe

for df in dfs:
    
#Converting columns alphabets to lowercase
    df.columns = [i.lower() for i in df.columns]
    
#Converting Date/Time column to a datastamp
    df['date/time'] = pd.to_datetime(df['date/time'], format="%m/%d/%Y %H:%M:%S")

#Create a new column for day name
    df['day_name'] = df['date/time'].dt.day_name()

#Create a new column for day
    df['day'] = df['date/time'].dt.day

#Create a new column for month
    df['month'] = df['date/time'].dt.month

#Create a new column for hour
    df['hour'] = df['date/time'].dt.hour

#Create a new column for minute
    df['minute'] = df['date/time'].dt.minute

uber_2014.head()
```

### Analysis of Location

```python
HeatMap(df.groupby(['lat','lon'])['day_name'].count().reset_index(),zoom=20,radius=15).add_to(basemap)
basemap
```

<b>Lets create a function for a specific day

```python
def plot_HeatMap(df,day):
    df_out=df[df['day_name']==day]
    df_out.groupby(['lat','lon'])['day_name'].count().reset_index()
    HeatMap(df_out.groupby(['lat','lon'])['day_name'].count().reset_index(),zoom=20,radius=15).add_to(basemap)
    return basemap
```

```python
plot_HeatMap(df,'Sunday')
```
