```python
def getPokemonTypes(pokemonNum):
    types = []
    type1=data['Type 1'][pokemonNum]
    type2=data['Type 2'][pokemonNum]
    types.append(typeDictionary[type1])
    if type2 in typeDictionary:
        types.append(typeDictionary[type2])
    return types

def getMatchupMultiplier(attackerTypes,defenderTypes):
    attackerMatchups=damageMatrix[attackerTypes]
    defenderSpecific=attackerMatchups[:,defenderTypes]
    # the attacker will hopefully choose their more damaging element
    result=np.max(np.prod(defenderSpecific,axis=1))
    return result

def calcHp(stat):
    return stat*2+110

def calcOtherStats(stat):
    return stat*2+5

def getPokemonStats(pokemonNum):
    stats=data[["HP","Attack","Defense","Sp. Atk","Sp. Def","Speed"]].to_numpy()
    hp, a, d, sa,sd,spd = stats[pokemonNum]
    return [calcHp(hp),
           calcOtherStats(a),
           calcOtherStats(d),
           calcOtherStats(sa),
           calcOtherStats(sd),
           calcOtherStats(spd)]

def damageCalculation(attack,defense,multiplier):
    # equation pulled from bulbapedia
    # https://bulbapedia.bulbagarden.net/wiki/Damage
    # some modifiers including the random one are ignored.
    level = 100 # the stats are calculated at lvl 100
    power = 80 # this is arbitrary, i feel 80 is a common move damage.
    stab = 1.5 # its assumed that pokemon are always using stab moves.
    return ((((2*level)/5+2)*power*attack/defense)/50+2)*multiplier*stab

def getDamageDelta(attacker,defender):
    aTypes,aStats = attacker
    dTypes,dStats = defender
    multiplier=getMatchupMultiplier(aTypes,dTypes)
    _, a, _, sa,_,_ = aStats
    _, _, d, _,sd,_ = dStats
    # requires damage to be 1 minimally
    damageDelta = np.max([damageCalculation(a,d,multiplier),
                    damageCalculation(sa,sd,multiplier),
                    1])
    return damageDelta
    

def fightPokemon(a,b): # returns 1 if a wins, 0 otherwise
    # calculate the amount of damage each pokemon would do to the other 
    # assuming they used a move of their more effective element 
    _,aStats = a
    _,bStats = b
    aHp = aStats[0]
    aSpd = aStats[5]
    bHp = bStats[0]
    bSpd = bStats[5]
    aDmgDelta = getDamageDelta(a,b)
    bDmgDelta = getDamageDelta(b,a)
    while(aHp>0 and bHp>0):
        aHp-=bDmgDelta
        bHp-=aDmgDelta
    if aHp<= 0 and bHp <=0:
        # then the faster pokemon wins
        # simulates them having moved first on the last turn.
        if(aSpd>bSpd):
            return 1
        else:
            return 0
    elif bHp<=0:
        return 1
    else: #aHp <= 0
        return 0
    

#process pokemon
pokemon = []
for i in range(800):
    types = getPokemonTypes(i)
    stats = getPokemonStats(i)
    pokemon.append((types,stats))

    
# test bulbasaur attacking 
print('multipliers')
bulbasaurTypes = getPokemonTypes(0)
charizardTypes = getPokemonTypes(8)
bvsc = getMatchupMultiplier(bulbasaurTypes,charizardTypes)
cvsb = getMatchupMultiplier(charizardTypes,bulbasaurTypes)
print('bulbasaur vs charizard: ',bvsc)
print('charizard vs bulbasaur: ',cvsb)

print('damage')
bulba = pokemon[0]
charz = pokemon[8]
print(getDamageDelta(bulba,charz))
print(getDamageDelta(charz,bulba))
fightResult = fightPokemon(charz,bulba)
print(fightResult)
print(bulba)
print(charz)
    
```

```python
resultGrid = np.zeros((800,800))
# fight all the pokemon
for ai in range(800):
    attacker = pokemon[ai]
    for di in range(800):
        defender = pokemon[di]
        result = fightPokemon(attacker,defender)
        resultGrid[ai,di] = result
print(resultGrid)
```

```python
victories = np.sum(resultGrid,axis=1)
```

```python
def getResultsByName(pokemonName):
    pokemonNum = getPokemonNumByName(pokemonName)
    return getResults(pokemonNum)
    
def getResults(pokemonNum):
    winResults = resultGrid[pokemonNum]
    won = []
    for i in range(len(winResults)):
        if winResults[i]==1:
            won.append(getPokemonName(i))
    return (victories[pokemonNum],won)
```

```python
print('most victories: ',np.max(victories))
print('by pokemon: ',getPokemonName(np.argmax(victories)))

print('least victories: ', np.min(victories))
print('by pokemon: ',getPokemonName(np.argmin(victories)))
```
