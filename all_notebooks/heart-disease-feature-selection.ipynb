```python
df = pd.read_csv('/kaggle/input/heart-disease-uci/heart.csv')
y_data = df.target.values
x_data = df.drop(['target'], axis = 1)
```

```python
# Information Gain

from sklearn.feature_selection import mutual_info_classif

importance = mutual_info_classif(x_data, y_data)

feature_importance_Ig = pd.Series(importance, index=x_data.columns)

fig,ax = plt.subplots()
fig.set_size_inches(15,6)
ax.barh(feature_importance_Ig.sort_values().index, feature_importance_Ig.sort_values().values)
plt.xticks(rotation=90)
plt.title('Information Gain for each feature')
plt.show()
```

```python
from skfeature.function.similarity_based import fisher_score
%matplotlib inline

ranks = fisher_score.fisher_score(X, y)

feat_importances = pd.Series(ranks, df.columns[0:len(df.columns)-1])
feat_importances.plot(kind="barh")
plt.show()
```

```python
# Variance Threshold

from sklearn.feature_selection import VarianceThreshold
v_threshold = VarianceThreshold(threshold=0.01)
v_threshold.fit(X)  # fit finds the features with zero variance
v_threshold.get_support()
```

```python
# MAD

mean_abs_diff = np.sum(np.abs(x_data- np.mean(x_data, axis=0)), axis=0)/x_data.shape[0]

plt.bar(np.arange(x_data.shape[1]),mean_abs_diff)
```

```python
# Dispersion Ratio

X = X +1 #avoid 0 division
aritmeticMean = np.mean(X, axis =0 )
geometricMean = np.power(np.prod(X, axis =0 ),1/X.shape[0])
R = aritmeticMean/geometricMean

plt.bar(np.arange(X.shape[1]),R)
```
