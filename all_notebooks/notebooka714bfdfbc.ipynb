### Process data

```python
def get_number(text):
    """
    remove the dollar and B symbols and then convert to float
    """
    try:
        text_to_number = float(text.replace("$","").replace(" B","").replace(",",""))
        
    except:
        text_to_number = float(text.replace("$","").replace(" M","").replace(",",""))/1000
                
    return text_to_number
    
```

```python
df1 = df.copy()
```

```python
for i in range(3,7):
    df1.iloc[:,i] = df1.iloc[:,i].apply(get_number)
df1.head(4)
```

```python
df1_melt = df1.melt(id_vars=["Rank", "Name","Country"], var_name="Parameter", value_name="Value")
df1_melt.head()
```

```python
df1['cut'] = pd.cut(df1['Rank'],bins=range(0,550,50),labels=range(10))
df1
```
