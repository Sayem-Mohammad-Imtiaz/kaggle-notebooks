```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfm=pd.read_csv("../input/skillsay-ai-crickethon-dataset/matches.csv")
dfb=pd.read_csv("../input/skillsay-ai-crickethon-dataset/deliveries.csv")
```

```python
dfm.info()
```

```python
dfb.info()
```

```python
dfm=dfm.replace(to_replace="Delhi Daredevils", value="Delhi Capitals")
dfb=dfb.replace(to_replace="Delhi Daredevils", value="Delhi Capitals")
```

```python
df=dfm.merge(dfb, left_on='id', right_on='match_id')
df
```

```python
latest=df[(df["season"]>2015)]
latest
```

```python
fig,ax=plt.subplots(figsize=(11,11))
df["winner"].value_counts().plot.pie(autopct='%.2f',shadow=True);
```

```python
fig,ax=plt.subplots(figsize=(11,11))
latest["winner"].value_counts().plot.pie(autopct='%.2f',shadow=True);
```

```python
sixes=df[(df["batsman_runs"]==6)]
sixes
```

Which player hit most sixes?

```python
fig,ax=plt.subplots(figsize=(20,10))
sns.countplot(y='batsman',
              data=sixes,
              palette="copper",
              order=sixes['batsman'].value_counts().iloc[:10].index)
plt.show()
```
