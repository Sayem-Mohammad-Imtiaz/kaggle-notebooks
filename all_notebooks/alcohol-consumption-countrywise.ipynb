```python
import math
import collections

import numpy as np
import pandas as pd
import matplotlib.pyplot as pp
import seaborn as sns

%matplotlib inline
```

```python
alcohol=pd.read_csv('../input/alcohol-consumption/gapminder_alcohol.csv')
```

```python
alcohol.head()
```

```python
alcohol.describe()
```

```python
#Checking nulls in data
ns=alcohol.isnull().sum()
ns
```

```python
#Creating copy of the original data
alcohol_original=alcohol.copy()
```

```python
#Filling nulls with mean values
alcohol['alcconsumption'].fillna(alcohol['alcconsumption'].mean(), inplace=True)
alcohol['incomeperperson'].fillna(alcohol['incomeperperson'].mean(), inplace=True)
alcohol['suicideper100th'].fillna(alcohol['suicideper100th'].mean(), inplace=True)
alcohol['employrate'].fillna(alcohol['employrate'].mean(), inplace=True)
alcohol['urbanrate'].fillna(alcohol['urbanrate'].mean(), inplace=True)
```

```python
#Checking nulls in data
ns=alcohol.isnull().sum()
ns
```

```python
#Plotting suicide rate against alcoholism:
pp.scatter(alcohol.alcconsumption, alcohol.suicideper100th)
pp.xlabel('Alcohol Counsumption')
pp.ylabel('Suicide Rate')
pp.title('Alcohol and Suicide rate')
pp.show()
```

```python
#Plotting suicide rate against employment rate
pp.scatter(alcohol.employrate, alcohol.suicideper100th)
pp.xlabel('Employment Rate')
pp.ylabel('Suicide Rate')
pp.title('Employment and Suicide rate')
pp.show()
```

```python
#Lowest alcohol consuming countries
pp.figure(figsize=(25,12))
top=sns.barplot(data=alcohol.sort_values('alcconsumption').head(10),x='country', y='alcconsumption')
pp.tick_params(labelsize=10)
pp.title('Bottom 10 countries by alcohol consumption')
```

```python
#Top 10 alcohol consuming countries

pp.figure(figsize=(25,12))
sns.barplot(data=alcohol.sort_values('alcconsumption').tail(10),x='country', y='alcconsumption')
pp.tick_params(labelsize=10)
pp.title('Top 10 countries by alcohol consumption')
```

```python
#Plotting income per person to Urban rate
y=np.array([alcohol['employrate'].min(),alcohol['employrate'].mean(),alcohol['employrate'].max()])
y1=np.array([alcohol['urbanrate'].min(),alcohol['urbanrate'].mean(),alcohol['urbanrate'].max()])
x=['Min', 'Avg', 'Max']
pp.bar(x,y,label='employrate')
pp.bar(x,y1,label='Urban Rate',bottom=y)
pp.title("Urbanisation and Employment rate around the world")
pp.xlabel('Level')
pp.ylabel('Urbanisation and Employment rate')
pp.legend()
pp.show()
```

```python
#Creating an index to get partiicular countries further
alcohol_index=alcohol.set_index("country")
```

```python
#Taking 10 developed and developing countries for comparison on different parmeters
arr=['India', 'United States', 'United Kingdom', 'Pakistan', 'Afghanistan', 'New Zealand', 'Sri Lanka', 'China', 'Norway', 'Nepal']
countries=alcohol_index.loc[arr]
countries

pp.figure(figsize=(25,12))
pp.subplot(2,2,1)
x=arr
y=countries['urbanrate']
pp.xlabel("Countries")
pp.ylabel("Urban rate")
pp.title ("Urbanisation")
pp.bar(x,y)
pp.axhline(countries['urbanrate'].mean(), linestyle=':')
pp.tick_params(labelsize=5)
pp.show()

pp.figure(figsize=(25,12))
pp.subplot(2,2,2)
y=countries['alcconsumption']
y1=countries['suicideper100th']
pp.plot(arr,y,label='alchohol consumption')
pp.plot(arr, y1,label='Suicide rate')
pp.xlabel("Countries")
pp.ylabel("Alcohol and Suicide")
pp.title ("Alcohol Consumption and Suicide rate")
pp.tick_params(labelsize=5)
pp.legend()
pp.show()

pp.figure(figsize=(25,15))
pp.subplot(2,2,3)
x=countries['incomeperperson']
pp.pie(x, labels=arr)
pp.legend(loc='right', bbox_to_anchor =(1, 0, 0.5, 1))
pp.title("Income/ person")
pp.tick_params(labelsize=5)
pp.show()
```
