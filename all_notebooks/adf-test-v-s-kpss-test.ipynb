```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_colwidth', -1)
```

```python
data=pd.read_csv('/kaggle/input/exchange-rate/exchange-rate-twi.csv')
data.columns = ['Months','Exchange Rate TWI']

data.head()
```

```python
from datetime import date
from dateutil.relativedelta import relativedelta
for i in range(0,len(data.Months)):
    data.Months[i] = date(1990,1,1) + relativedelta(months=+i)

data.head()
```

```python
data = data.set_index('Months')
data.head()
```

```python
data.plot(figsize=(20, 4))
plt.grid()
plt.legend(loc='best')
plt.title('Exchange Rate TWI')
plt.show(block=False)
```

# Augmented Dickey-Fuller (ADF) test

```python
from statsmodels.tsa.stattools import adfuller
adf_test = adfuller(data['Exchange Rate TWI'])

print('ADF Statistic: %f' % adf_test[0])
print('Critical Values @ 0.05: %.2f' % adf_test[4]['5%'])
print('p-value: %f' % adf_test[1])
```

Failed to reject Null Hypothesis, The series is not stationary

# Kwiatkowski-Phillips-Schmidt-Shin (KPSS) test

```python
from statsmodels.tsa.stattools import kpss
kpss_test = kpss(data['Exchange Rate TWI'])

print('KPSS Statistic: %f' % kpss_test[0])
print('Critical Values @ 0.05: %.2f' % kpss_test[3]['5%'])
print('p-value: %f' % kpss_test[1])
```

Reject Null Hypothesis, The series is not stationary
