```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

#plotly
import plotly.graph_objects as go
import plotly.figure_factory as ff

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# Any results you write to the current directory are saved as output.
```

```python
df=pd.read_csv('/kaggle/input/coronavirus-covid19-dataset-india/corona_India_latest.csv')
```

```python
df.drop(["Unnamed: 0"],axis=1,inplace=True)
```

```python
df
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=df["Name of State / UT"],
    y=df["Total Confirmed cases"],
    name='TotalCases-Indian National',
    marker_color='#636EFA',
    text=df["Total Confirmed cases"],
    textposition='auto'
))

fig.add_trace(go.Bar(
    x=df["Name of State / UT"],
    y=df["Death"],
    name='Death',
    marker_color='#EF553B',
    text=df["Death"],
    textposition='auto'
))
fig.update_layout(title_text='Corona virus TotalCase in India',xaxis_tickfont_size=14,
    yaxis=dict(
        title='COUNT',
        titlefont_size=16,
        tickfont_size=14))
fig.show()


```

```python
df1=df.copy()
```

```python
df1.loc['total'] = df1.select_dtypes(np.number).sum()
df1
```

```python
fig = go.Figure()
data=[go.Bar(
    x=df["Name of State / UT"],
    y=df["Total Confirmed cases"],
    name='Total Confirmed cases',
    text=df["Total Confirmed cases"],
    textposition='auto',
    marker={
        'color': df["Total Confirmed cases"],
        'colorscale': 'balance'
    }
)]

layout = {
    'xaxis': {
        'categoryorder': 'array',
        'categoryarray': [x for _, x in sorted(zip(df1["Total Confirmed cases"], df1["Name of State / UT"]))]
    }
}

fig = go.FigureWidget(data=data, layout=layout)

fig.update_layout(title_text=("Total people affected due to Corona virus :"+str(df1.loc["total"]["Total Confirmed cases"])),xaxis_tickfont_size=14,
    yaxis=dict(
        title='COUNT',
        titlefont_size=16,
        tickfont_size=14))
fig.show()


```

```python
fig = go.Figure()
data=[go.Bar(
    x=df["Name of State / UT"],
    y=df["Death"],
    name='Death',
    text=df["Death"],
    textposition='auto',
    marker={
        'color': df["Death"],
        'colorscale': 'balance'
    }
)]

layout = {
    'xaxis': {
        'categoryorder': 'array',
        'categoryarray': [x for _, x in sorted(zip(df1["Death"], df1["Name of State / UT"]))]
    }
}

fig = go.FigureWidget(data=data, layout=layout)

fig.update_layout(title_text=("Total people died due to Corona virus :"+str(df1.loc["total"]["Death"])),xaxis_tickfont_size=14,
    yaxis=dict(
        title='COUNT',
        titlefont_size=16,
        tickfont_size=14))
fig.show()


```

```python
fig = go.Figure()
data=[go.Bar(
    x=df["Name of State / UT"],
    y=df["Active Cases"],
    name='Active Cases',
    text=df["Active Cases"],
    textposition='auto',
    marker={
        'color': df["Active Cases"],
        'colorscale': 'balance'
    }
)]

layout = {
    'xaxis': {
        'categoryorder': 'array',
        'categoryarray': [x for _, x in sorted(zip(df1["Active Cases"], df1["Name of State / UT"]))]
    }
}

fig = go.FigureWidget(data=data, layout=layout)

fig.update_layout(title_text=("Total No of Active Cases :"+str(df1.loc["total"]["Active Cases"])),xaxis_tickfont_size=14,
    yaxis=dict(
        title='COUNT',
        titlefont_size=16,
        tickfont_size=14))
fig.show()


```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=df["Name of State / UT"][:10],
    y=df["Total Confirmed cases"][:10],
    name='TotalCases: '+str(df1.loc["total"]["Total Confirmed cases"]),
    marker_color='#636EFA',
    text=df["Total Confirmed cases"][:10],
    textposition='auto'
))


fig.add_trace(go.Bar(
    x=df["Name of State / UT"][:10],
    y=df["Active Cases"][:10],
    name='Active Cases: '+str(df1.loc["total"]["Active Cases"]),
    marker_color='LightSkyBlue',
    text=df["Active Cases"][:10],
    textposition='auto'
))

fig.add_trace(go.Bar(
    x=df["Name of State / UT"][:10],
    y=df["Death"][:10],
    name='Death: '+str(df1.loc["total"]["Death"]),
    marker_color='#EF553B',
    text=df["Death"][:10],
    textposition='auto'
))
fig.update_layout(title_text='Top 10 affected states in India',xaxis_tickfont_size=14,
    yaxis=dict(
        title='COUNT',
        titlefont_size=16,
        tickfont_size=14))
fig.show()


```

```python
df3=df.copy()
```

```python
df3['D/C Ratio in percentage']=(df3['Death']/df3['Total Confirmed cases']*100).round(2)
```

```python
df3
```

```python
df3.sort_values("D/C Ratio in percentage", axis = 0, ascending = False, 
                 inplace = True, na_position ='last')
df3
```

```python
fig = go.Figure()
data=[go.Bar(
    x=df3["Name of State / UT"],
    y=df3["D/C Ratio in percentage"][df3["D/C Ratio in percentage"]>0],
    name='D/C Ratio in percentage',
    text=df3["D/C Ratio in percentage"][df3["D/C Ratio in percentage"]>0],
    textposition='auto',
    marker={
        'color': df3["D/C Ratio in percentage"],
        'colorscale': 'balance'
    }
)]

layout = {
    'xaxis': {
        'categoryorder': 'array',
        'categoryarray': [x for _, x in sorted(zip(df3["D/C Ratio in percentage"], df3["Name of State / UT"]))]
    }
}

fig = go.FigureWidget(data=data, layout=layout)

fig.update_layout(title_text=("Death per Confirmed Cases in terms of percentage "),xaxis_tickfont_size=14,
    yaxis=dict(
        title='COUNT',
        titlefont_size=16,
        tickfont_size=14))
fig.show()


```
