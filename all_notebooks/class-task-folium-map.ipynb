```python
data = pd.read_csv("/kaggle/input/pakistan-corona-virus-citywise-data/PK COVID-19.csv")
coords = pd.read_csv("/kaggle/input/pakistan-corona-virus-citywise-data/covid_cases_city_coordinates.csv")
coords['City'] == 'Abbottabad'
```

```python
cities = list(set(data['City'].values))
data.dropna()
#
city_cases = pd.DataFrame(columns=['City', 'Cases', 'Longitude', 'Latitude'])
for city in cities:
    cases =0
    for i,row in data.iterrows():
#         print(row['Cases'])
        if row['City'] == city:
            cases+=row['Cases']
            
    lat =coords[coords['City'] == city]['Latitude (generated)']
    lng = coords[coords['City'] == city]['Longitude (generated)']
    df = { "City": city, "Cases": cases, "Longitude": lng, "Latitude": lat}
#     print(df)
    city_cases=city_cases.append(pd.DataFrame(df))
```

```python
# calculating total cases
full_cases = 0
for i, row in city_cases.iterrows():
    full_cases+=row['Cases']
```

```python
# coords[coords['City'] == 'Abbottabad']
city_cases = city_cases.dropna()
```

```python
pakistan_map = folium.Map(location = [30.37,69.34],zoom_start=4.5,tiles='OpenStreetMap')
for city, cases, lng, lat in zip(list(city_cases['City']), list(city_cases['Cases']), list(city_cases['Longitude']), list(city_cases['Latitude'])):
    folium.CircleMarker(location=[lat, lng], radius=(float(cases)/full_cases)*100., color="purple", fill=True, fill_color="red").add_to(pakistan_map)
```
