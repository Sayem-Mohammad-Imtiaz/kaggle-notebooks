```python
import pandas as pd
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix

import seaborn as sns
```

```python
df=pd.read_csv('/kaggle/input/heart-disease-uci/heart.csv')
df
```

```python
df.info()
df.columns
```

```python

#for detecting and removing outliers
factor = 3
upper_lim = df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']].mean () + df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']].std () * factor
lower_lim = df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']].mean () - df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']].std () * factor

data = df[(df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']] < upper_lim) & (df[['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']] > lower_lim)]
```

```python
df.corr()
```

```python
x=df.drop(['target','sex','thal','exang'],axis=1)
y=df.target
```

```python
scaler = StandardScaler()
print(scaler.fit(x))
```

```python
pd.DataFrame(scaler.transform(x))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4)
```

```python
model=RandomForestClassifier(n_estimators=100)
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
```

```python
accuracy_score(y_test,predict)
```

```python
print(classification_report(y_test,predict))
```

```python
cm=confusion_matrix(y_test,predict)
sns.heatmap(cm,annot=True)
```
