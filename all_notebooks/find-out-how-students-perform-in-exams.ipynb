# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
```

```python
df.shape#dimension
```

```python
df.isnull().sum()#checking for missing values
```

```python
df.describe(include='all')#statistical summary
```

```python
df.head()
```

# Male vs Female ratio

```python
df['gender'].value_counts().plot(kind='pie',autopct='%1.1f%%',figsize=(6,6),colors=['mediumorchid', 'cornflowerblue'],explode=[0,0.04],startangle=90,shadow=True)
plt.title('Male vs Female ratio')
plt.show()
```

# Ethnicity ratio

```python
df['race/ethnicity'].value_counts().plot(kind='pie',autopct='%1.1f%%',figsize=(6,6),colors=['lightsteelblue', 'royalblue', 'lavender', 'navy', 'mediumblue'],startangle=90,explode=[0.1,0,0,0,0],shadow=True)
plt.title('Ethnicity ratio')
plt.show()
```

# Parents education level ratio

```python
df['parental level of education'].value_counts().plot(kind='pie',autopct='%1.1f%%',figsize=(8,8),colors=['rosybrown', 'indianred', 'firebrick', 'darkred', 'mistyrose','tomato'],explode=[0.04,0.04,0.04,0,0,0],shadow=True,startangle=90)
plt.title('Parents education level ratio')
plt.show()
```

# Top math scorers wrt ethnicity

```python
df.groupby('race/ethnicity')['math score'].max().sort_values(ascending=False).plot(kind='barh',figsize=(10,6),color='cyan')
plt.title('Top math scorers wrt ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('Math score')
plt.show()
```

# Top reading scorers wrt ethnicity

```python
df.groupby('race/ethnicity')['reading score'].max().sort_values(ascending=False).plot(kind='barh',figsize=(10,6),color='purple')
plt.title('Top reading scorers wrt ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('Math score')
plt.show()
```

# Top writing scorers wrt ethnicity

```python
df.groupby('race/ethnicity')['writing score'].max().sort_values(ascending=False).plot(kind='barh',figsize=(10,6),color='gray')
plt.title('Top writing scorers wrt ethnicity')
plt.xlabel('Ethnicity')
plt.ylabel('Math score')
plt.show()
```

from all the visualizations we can see that students from D ethnicity scores highets in every subject
