**THIS MODEL BELOW USES THE OPEN SOURCE OF FACEBOOK WHICH IS FB PROPHET TO PREDICT**

BELOW IS THE PREDICTION FOR HOW MANY NEW CAMES CAN OCCUR FOR THE NEXT 30 DAYS IN THE UNITED KINGDOM ACCODDING TO THE DATASET. 
NOTE: THE PREDICTION IS NOT REAL, IT IS JUST A POSSIBILITY.

```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from fbprophet import Prophet # open source by Facebook
from fbprophet.plot import plot_plotly, plot_components_plotly # part of fbprophet for plotting

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# storing dataset in variable df in reading with pandas
df = pd.read_csv('../input/united-kingdom-covid19-figures-aug-21/Cases by date reported United KIngdom.csv')
# checking our data
df.head(30)
```

```python
# column names change
df['ds'] = df['date']
df['y'] = df['newCasesByPublishDate']

# drop extra columns
df = df[['ds', 'y']]
```

```python
# making the model
m = Prophet()
# fitting our data in the model made
m.fit(df)
```

```python
# making future dataframe here the dates and new rows
future = m.make_future_dataframe(periods=30)
future.tail(30)
```

```python
# forecasting/predicting for future
forecast = m.predict(future)
# ds is the dates in format YYYY/MM/DD
# yhat is prediction
# yhat_lower is the lowest prediction possible
# yhat_upper is the prediction that can be most possible
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(30)
```

```python
# plotting the forecast
fig1 = m.plot(forecast)
```

```python
# checking components in data
fig2 = m.plot_components(forecast)
```

```python
# creating interctive figures of the forecast
plot_plotly(m, forecast)
```

```python
plot_components_plotly(m, forecast)
```
