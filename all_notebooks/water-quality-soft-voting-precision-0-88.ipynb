```python
water_data_drop=water_data.dropna()
water_data_drop
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
water_data_drop.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
np.random.seed(25)
water_data_drop_shuffled = water_data_drop.sample(frac=1)
```

```python
water_data_drop_shuffled
```

```python
X_full=water_data_drop_shuffled.drop("Potability", axis=1)
y_full=water_data_drop_shuffled["Potability"].copy()
X_full,y_full
```

```python
X_full.corr()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_full=sc.fit_transform(X_full)
```

```python
index=len(X_full)
X_train = X_full[:int(index*0.8)]
X_test = X_full[int(index*0.8):]

y_train = y_full[:int(index*0.8)]
y_test = y_full[int(index*0.8):]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import GradientBoostingClassifier

log_clf = LogisticRegression(solver="liblinear", random_state=25)
rnd_clf = RandomForestClassifier(n_estimators=500, random_state=25)
svm_clf = SVC(kernel="rbf",gamma="scale", probability=True,class_weight="balanced", random_state=25)
gb_clf = GradientBoostingClassifier(n_estimators=500, learning_rate=0.01,max_depth=10, random_state=25)

voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='soft')
```

```python
voting_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
con_matrix=confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import precision_score
precision_score(y_test, y_pred)
```
