RAM ve OnKameraMP değişkenlerinde eksik verilerimizin olduğunu görmüştük. Eksik verileri bu değişkenlerin ortalamaları ile dolduralım. Sonrasında da eksik verimiz var mı diye tekrardan kontrol edelim.



```python
df['RAM'] = df['RAM'].fillna((df['RAM'].mean()))
df['OnKameraMP'] = df['OnKameraMP'].fillna((df['OnKameraMP'].mean()))
df.isna().sum()#Eksik verilerden kurtulduk.
```

FiyatAraligi, Renk, Bluetooth, CiftHat, 4G, 3G, Dokunmatik, WiFi değişkenlerimiz sayısal değildir. Bundan dolayı bu değişkenlerimizi sayısal yapmamız gerekir.

```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
df["FiyatAraligi1"] = lb_make.fit_transform(df["FiyatAraligi"])

lb_make = LabelEncoder()
df["Renk1"] = lb_make.fit_transform(df["Renk"])

lb_make = LabelEncoder()
df["Bluetooth1"] = lb_make.fit_transform(df["Bluetooth"])

lb_make = LabelEncoder()
df["CiftHat1"] = lb_make.fit_transform(df["CiftHat"])

lb_make = LabelEncoder()
df["Internet4G"] = lb_make.fit_transform(df["4G"])

lb_make = LabelEncoder()
df["Internet3G"] = lb_make.fit_transform(df["3G"])

lb_make = LabelEncoder()
df["Dokunmatik1"] = lb_make.fit_transform(df["Dokunmatik"])

lb_make = LabelEncoder()
df["WiFi1"] = lb_make.fit_transform(df["WiFi"])




df.FiyatAraligi1=df["FiyatAraligi1"].astype("category")
df.Renk1=df["Renk1"].astype("category")
df.Bluetooth1=df["Bluetooth1"].astype("category")
df.CiftHat1=df["CiftHat1"].astype("category")
df.Internet4G=df["Internet4G"].astype("category")
df.Internet3G=df["Internet3G"].astype("category")
df.Dokunmatik1=df["Dokunmatik1"].astype("category")
df.WiFi1=df["WiFi1"].astype("category")



Fiyatlar= df[["FiyatAraligi", "FiyatAraligi1"]]
Renkler= df[["Renk", "Renk1"]]
Bluetooth1= df[["Bluetooth", "Bluetooth1"]]
CiftHat1= df[["CiftHat", "CiftHat1"]]
Internet4G= df[["4G", "Internet4G"]]
Internet3G= df[["3G", "Internet3G"]]
Dokunmatik1= df[["Dokunmatik", "Dokunmatik1"]]
WiFi1= df[["WiFi", "WiFi1"]]



del df["FiyatAraligi"]
del df["Renk"]
del df["Bluetooth"]
del df["CiftHat"]
del df["4G"]
del df["3G"]
del df["Dokunmatik"]
del df["WiFi"]
```

# Bu bağımlı ve bağımsız değişkenlerden train ve test olmak üzere 4 parça değişken oluşturalım. Bunu yapmak için train_test_split kullanalım. %25 oranda test istediğimiz için test_size=1/4 yapalım.
# 


```python
from sklearn.model_selection import train_test_split, GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/4, random_state = 3, shuffle = 1)
```

GaussianNB'u çekirdeğe dahil edip modelimizi inşa edelim. Sonrasında modelimizi eğitelim ve test edelim.



```python
from sklearn.naive_bayes import GaussianNB
gaussian = GaussianNB()
model_gaussian = gaussian.fit(X_train, y_train)
y_predict = gaussian.predict(X_test)
```

```python
print(classification_report(y_test,y_predict))
```

# DecisionTreeClassifier'ı çekirdeğe dahil edip modelimizi inşa edelim. Parametre olarak gini kullanalım. Sonrasında modelimizi eğitelim ve test edelim.

```python
from sklearn.tree import DecisionTreeClassifier
dtc_gini= DecisionTreeClassifier(criterion = 'gini')
model_dtc_gini = dtc_gini.fit(X_train,y_train)
y_predict = dtc_gini.predict(X_test)
```

```python
print(classification_report(y_test,y_predict))
```

Bu sefer parametre olarak entropy kullanalım. DecisionTreeClassifier'ı zaten çekirdeğe dahil etmiştik bu yüzden direk modelimizi oluşturabiliriz. Oluşturduktan sonra eğitelim ve test edelim.

```python
dtc_entropy= DecisionTreeClassifier(criterion = 'entropy')
model_dtc_entropy = dtc_entropy.fit(X_train,y_train)
y_predict = dtc_entropy.predict(X_test)
```

```python
print(classification_report(y_test,y_predict))
```

KNeighborsClassifier'ı çekirdeğe dahil edip modelimizi inşa edelim. Sonrasında modelimizi eğitelim ve test edelim.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
model_knn= knn.fit(X_train, y_train)
y_predict = knn.predict(X_test)
```

```python
print(classification_report(y_test,y_predict))
```

KNN modelinde komşu sayısını 2'den 15'e kadar alalım ve plot grafiğimizi çizelim

```python
score_list = []
for each in range(2,15):
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(X_train,y_train)
    score_list.append(knn2.score(X_test, y_test))

plt.plot(range(2,15),score_list)
plt.xlabel("k en yakın komşu sayıları")
plt.ylabel("doğruluk skoru")
plt.show()
```
