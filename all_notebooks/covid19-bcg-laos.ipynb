The country's National Task Force for COVID-19 Prevention and Control announced at a press conference that more offices and businesses will resume normal operations. Some schools will reopen and certain sports activities will be permitted, reports The Laotian Times. 

Laos' 7 million citizens will also be allowed to travel domestically, though foreign travelers are still not allowed in.
