```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#check data
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#load dataset
local_frame = pd.read_csv('/kaggle/input/dogecoin-top-daily-reddit-submission/doge_12htop100_DailySub_0101_to_0710_PushShift.csv')
#convert 'score' column to float
local_frame['score'] = local_frame['score'].astype(float) 
local_frame.head()
```

```python
#showing the 25 subreddits that appear the most
local_frame['subreddit'].value_counts()[:25].plot.bar()
plt.show()

print('\nIgnoring the dogecoin subreddit for a better visualization\n')
local_frame['subreddit'].value_counts()[1:25].plot.bar()
plt.show()
```

```python
#grouping by number of comments per day
total_num_comments = local_frame.groupby('date')['num_comments'].sum()
#showing the 20 dates with the most comments
print(total_num_comments.sort_values(ascending=False)[:20])
total_num_comments.sort_values(ascending=False)[:20].plot.bar()
```

```python
#grouping by total score per day
score_per_day = local_frame.groupby('date')['score'].sum()
print(score_per_day.sort_values(ascending=False))
score_per_day.plot()
```

```python
#showing the 10 dates with the most upvotes
print(score_per_day.sort_values(ascending=False)[:10])
score_per_day.sort_values(ascending=False)[:10].plot.bar()
```
