**Part I:**
Consider a pandas series which is indexed by integers and whose elements are integers as well. Create a series consisting of those elements which are indexed by indices which are divisible by 3.

```python

def exI(A):
    V = A.index
    U = V[V % 3 == 0]
    return A[U]
```

```python
A = pd.Series(np.array([10,11,12,3,4,5,7,9,15,20]),index = [10,21,33,36,40,45,52,61,76,96])
# or we can use A1 = A[(A.index%3 ==0)]
```

```python
exI(A)
```

**Part II:** 
Car prices prediction

```python
# import the data.
df = pd.read_csv('/kaggle/input/datasetexam3/Group3.csv')
```

```python
df.isna().any().any()
```

```python
df.head()
```

```python
df.info()
```

```python
df.mean()
```

```python
df1 = df.fillna(df.mean()) # Fill out missing values by using mean instead.
```

```python
df1
```

```python
columnlist = ['Year','Present_Price','Kms_Driven','Owner']
x = df1[columnlist]
y = df1['Selling_Price']
```

```python
x.head()
```

```python
# X - the columns to use for prediction.
from sklearn.model_selection import train_test_split
```

```python
# Create data for training and testing, leave 20% of data for testing.
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

**linear regression**

```python
from sklearn.linear_model import LinearRegression 
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
ypred = lr.predict(x_test)
ypred_train = lr.predict(x_train)
```

```python
# Error tolerance
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
mean_absolute_error(ypred_train,y_train)
```

```python
mean_absolute_error(ypred,y_test)
```

```python
r2_score(ypred,y_test)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(x_test.shape[0]),y_test,'r',range(x_test.shape[0]),ypred,'b')
plt.legend(['Actual price','Predicted price'])
```

**Decision tree regression**

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```

```python
ypred_dt = dt.predict(x_test)
ypred_traindt = dt.predict(x_train)
```

```python
mean_absolute_error(ypred_traindt,y_train)
```

```python
mean_absolute_error(ypred_dt,y_test)
```

```python
r2_score(ypred_traindt,y_train)
```

```python
r2_score(ypred_dt,y_test)
```

**Support vector machine**

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(y_test.index,y_test,'-r',y_test.index,ypred,'-b')
```

```python
from sklearn.svm import SVR
```

```python
svr = SVR()
svr.fit(x_train,y_train)
ypred_SVM = svr.predict(x_test)
```

```python
mean_absolute_error(ypred_SVM,y_test) / y_test.abs().mean()
```

**Neural networks**

```python
from sklearn.neural_network import MLPRegressor
```

```python
nr = MLPRegressor(hidden_layer_sizes = 3*(10,10,10,10))
```

```python
nr.fit(x_train,y_train)
```

```python
ypredNN = nr.predict(x_test)
```

```python
mean_squared_error(ypredNN,y_test)
```
