```python
import seaborn as sns
```

```python
sns.lmplot(x="crim",y="medv",data=boston_housing)
```

```python
sns.distplot(boston_housing["tax"])
```

```python
sns.distplot(boston_housing["rm"],kde=True)
```

```python
sns.distplot(boston_housing["medv"],bins=100)
```

```python
sns.boxplot(boston_housing["medv"],orient = "v")
```

```python
import math
```

```python
 boston_housing["rm_int"] = boston_housing["rm"].map(math.floor) 
```

```python
sns.boxplot(x="rm_int", y = "medv",data = boston_housing, orient="v")
```

```python
sns.boxplot(x="rm_int",y="medv",hue="chas",data = boston_housing, orient="v")
```

```python
sns.countplot(x = "rm_int", data = boston_housing)
```

```python
sns.countplot(x="rm_int",hue="chas",data=boston_housing)
```

```python
sns.pairplot(boston_housing[["crim","nox","rm","lstat","medv"]])
```

```python
import matplotlib.pyplot as plt 
```

```python
corr = boston_housing.corr()#特征的相关系数矩阵 
f, ax = plt.subplots(figsize=(10, 10)) 
cmap = sns.diverging_palette(220, 10, as_cmap=True) 
sns.heatmap(corr, cmap=cmap, vmax=1.0, square=True, xticklabels=2, yticklabels=2, linewidths=.3, cbar_kws={"shrink": .5}, ax=ax) 
plt.show()
```
