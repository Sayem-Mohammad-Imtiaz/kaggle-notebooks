**We Need to convert classification values to numerical values**

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#Create Object from LabelEncoder
label_En = LabelEncoder()
```

```python
data_PlayTennis['outlook']= label_En.fit_transform(data_PlayTennis['outlook'])
data_PlayTennis['temp']=label_En.fit_transform(data_PlayTennis['temp'])
data_PlayTennis['humidity']=label_En.fit_transform(data_PlayTennis['humidity'])
data_PlayTennis['windy']=label_En.fit_transform(data_PlayTennis['windy'])
data_PlayTennis['play']=label_En.fit_transform(data_PlayTennis['play'])
```

```python
data_PlayTennis
```

```python
data_PlayTennis.head()
```

```python
print(data_PlayTennis.columns)
```

```python
X = data_PlayTennis.drop(['play'] , axis=1 )
Y = data_PlayTennis['play']
```

```python
print('The Features is ::\n' , X)
```

```python
print('The Target Is (Play)::\n' , Y)
```

```python
#Importing train_test_split
from sklearn.model_selection import train_test_split
x_train  , x_test, y_train, y_test =train_test_split(X ,Y , test_size=0.3 ,stratify=Y ,random_state=101)
```

# Importing Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#Create Object from Decision Tree Classifier
D_T_C_Model =DecisionTreeClassifier(criterion='entropy' ,random_state=10)
```

```python
D_T_C_Model.fit(x_train , y_train)
```

```python
#check training score
D_T_C_Model.score(x_train , y_train)
```

```python
#check testing score
D_T_C_Model.score(x_test , y_test)
```

```python
y_pred = D_T_C_Model.predict(x_test)
y_pred
```

```python
D_T_C_Model.predict_proba(x_test)
```

```python
#visualize the tree using tree.plot_tree
from sklearn import tree
tree.plot_tree(D_T_C_Model)
```
