**Import the necessary library**

```python
#import necessary library
import numpy as np 
import pandas as pd
import os
import seaborn , matplotlib.pyplot as plt
```

**Read Data**

```python
#read dataset
data=pd.read_csv("/kaggle/input/high-quality-singers/Project-3 - Merged Datasheet.csv")
data.head()
```

**Unique Artist Per Year**

```python
#artist per year
data_2=data.filter(items=['Artist',"Year"])
data_2=data_2.groupby(["Year"])
data_2=data_2.describe()
plt.plot(data_2.Artist.unique,color='green', marker='o', linestyle='dashed',linewidth=2, markersize=12)
plt.ylabel("Number of Unique Artist")
plt.xlabel("Year")
plt.title("Unique Artist Per Year")
plt.show()
```

```python
#check shape
print(data.shape)
```

```python
#number of artist
len(np.unique(data.Artist))
```

**Most Song by Artist**

```python
#most song by the artist in dataset
print("Most Song by the Artist:",data.describe(include='O').Artist.top)
print("Number of Song by the Artist:",data.describe(include='O').Artist.freq)
```

**Year and the Best Artist List**

```python
data_3=data.filter(['Year',"Artist","Score"])
data_3=data_3[data_3.Score==100]
data_3=data_3.set_index("Year")
print("Best singer in year by score:\n",data_3.head(len(data.Year)))
```

```python
#performance of Artist (Whitney Houstan) in each year 
from matplotlib.pyplot import figure

figure(figsize=(8, 6), dpi=80)

data_4=data.filter(['Year','Artist',"Score"])
data_4=data_4[data_4.Artist=='Whitney Houston']

bars = data_4.Year.values
x_pos = np.arange(len(bars))
plt.xticks(x_pos, bars)

plt.bar(x_pos,data_4.Score,color='green')



plt.ylabel("Year")
plt.xlabel("Score")
plt.title("Performance of Whitney Houston in years")
plt.show()
```

```python
#Rank of Artist (Whitney Houstan) in each year 
from matplotlib.pyplot import figure

figure(figsize=(8, 6), dpi=80)

data_4=data.filter(['Year','Artist',"Rank"])
data_4=data_4[data_4.Artist=='Whitney Houston']

bars = data_4.Year.values

plt.pie(data_4.Rank,labels=bars,wedgeprops = { 'linewidth' : 3, 'edgecolor' : 'white' })
plt.title("Rank of Whitney Houston in years")
plt.suptitle("Lesser the Area ,Better the Rank", fontsize=8)
plt.show()
```
