```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats
```

```python
car=pd.read_csv("../input/vehicleprice/Data Analysis Challenge Data Set.csv")
```

```python
car.info()
```

```python
car.describe()
```

```python
plt.figure(figsize=(14,8))
sns.countplot(data=car,x="fuel-type")
plt.title("TYPE OF FUEL USED")
```

```python
order1=car["make"].value_counts().index
plt.figure(figsize=(25,8))
sns.countplot(data=car,x="make",order=order1)
```

```python
plt.figure(figsize=(25,8))
sns.countplot(data=car,x="make",hue="fuel-type",order=order1)
plt.xlabel("COMPANY NAME")
plt.ylabel("NO OF VEHICLE")
```

```python
plt.figure(figsize=(14,8))
order2=car["body-style"].value_counts().index
sns.countplot(data=car,x="body-style",order=order2)
plt.xlabel("BODY STYLE")
plt.ylabel("NO OF VEHICLE")
plt.title("BODY STYLES")
```

```python
plt.figure(figsize=(14,8))
order3=car["engine-location"].value_counts().index
sns.countplot(data=car,x="engine-location",order=order3)
plt.xlabel("engine-location")
plt.ylabel("NO OF VEHICLE")
plt.title("engine-location")
```

```python
plt.figure(figsize=(14,8))
order3=car["drive-wheels"].value_counts().index
sns.countplot(data=car,x="drive-wheels",order=order3)
plt.xlabel("drive-wheels")
plt.ylabel("NO OF VEHICLE")
plt.title("drive-wheels")
```

```python
order4=car["num-of-doors"].value_counts().index
sns.countplot(data=car,x="num-of-doors",order=order4)
```
