```python
#analise com 40 alunos
#recorte na base de dados
altura1=altura[0:40]
```

```python
altura1.info
```

```python
#medidas discritivas
altura1.describe()
```

```python
#importar biblioteca de graficos phyton
import matplotlib as plt
import seaborn as sns
```

```python
sns.boxplot(x="Altura",data=altura1)
```

```python
sns.distplot(altura1.dropna(), label='altura dos alunos')
```

Quando se retira 6 alunos, os dados ficam mais bem distribuídos, tornando-o mais simétrico. Além disso, não há mais a presença de outlier, deixando a calda menor, com uma menor amplitude dos dados. 

```python
#analise comparativa
sns.boxplot(x="Altura",data=altura1)
sns.boxplot(x="Altura",data=altura)
```

```python
#analise comparativa
sns.distplot(altura1.dropna(), label='altura dos alunos')
sns.distplot(altura.dropna(), label='altura dos alunos')
```

Comparadando as duas análises, percebe-se que com a retirada dos 6 alunos da distribuição, o gráfico ficou mais simétrico, com uma distribuição menos dispersa. Com a retirada desses alunos, a média e a mediana não teve uma teve uma mudança significativa.
