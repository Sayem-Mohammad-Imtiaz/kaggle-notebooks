```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=my_data['key'], y= my_data['loudness'], hue=my_data['genre'], legend=False)
plt.legend(loc='center right', bbox_to_anchor=(1.25, 0.5))
```

```python
df2 = my_data.dropna(axis=1)
df2 = df2.drop(labels=["type", "id", "uri", "track_href", "analysis_url"], axis=1)
df2.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df2['genre'], y=df2['tempo'])
plt.xticks(rotation=45)
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=df2['danceability'], y=df2['liveness'], hue=df2['genre'])
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(x=df2['instrumentalness'], y=df2['speechiness'], hue=df2['genre'])
```

```python
df3 = df2.drop(labels='genre', axis=1)
labels = df2['genre']
df3.head()
```
