```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv("/kaggle/input/students-performance-in-exams/StudentsPerformance.csv")
```

```python
df
```

```python
#general info
df.info()
```

```python
df.describe()
```

```python
#students that have had test preparation course
df_prep = df[df["test preparation course"]=="completed"]
#students that did not have any preparation course
df_noprep = df[df["test preparation course"]=="none"]
```

```python
#math score
sns.distplot(df_prep["math score"])
sns.distplot(df_noprep["math score"])
from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color="blue", lw=4),Line2D([0], [0], color="orange", lw=4)]
plt.legend(custom_lines,["Attended","Did not attend"])
plt.title("With and W/o preparation course - Math score")
```

```python
#reading score
sns.distplot(df_prep["reading score"])
sns.distplot(df_noprep["reading score"])
from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color="blue", lw=4),Line2D([0], [0], color="orange", lw=4)]
plt.legend(custom_lines,["Attended","Did not attend"])
plt.title("With and W/o preparation course - Reading score")
```

```python
#writing score
sns.distplot(df_prep["writing score"])
sns.distplot(df_noprep["writing score"])
from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color="blue", lw=4),Line2D([0], [0], color="orange", lw=4)]
plt.legend(custom_lines,["Attended","Did not attend"])
plt.title("With and W/o preparation course - Writing score")
```

#Students who attended a preparation course seems to be doing better than students who did not have any preparation courses.
The difference in score can be more easily seen in reading and writing score than math score.

```python
#student average score
df["average score"] = (df["math score"]+df["reading score"]+df["writing score"])/3
```

```python
#distribution of average score between male and female
sns.distplot(df[df["gender"]=="female"]["average score"],color = "red")
sns.distplot(df[df["gender"]=="male"]["average score"],color = "blue")
custom_lines = [Line2D([0], [0], color="red", lw=4),Line2D([0], [0], color="blue", lw=4)]
plt.legend(custom_lines,["female","male"])
plt.title("Average score by gender")
```

```python
#density plot of average score by lunch type
sns.distplot(df[df["lunch"]=="standard"]["average score"],color = "red")
sns.distplot(df[df["lunch"]=="free/reduced"]["average score"],color = "blue")
custom_lines = [Line2D([0], [0], color="red", lw=4),Line2D([0], [0], color="blue", lw=4)]
plt.legend(custom_lines,["standard lunch","free/reduced lunch"])
plt.title("Average score by lunch type")
```

```python
#density plot of average score by parent's education level
sns.distplot(df[df["parental level of education"]==("high school"or"some high school")]["average score"],color = "red")
sns.distplot(df[df["parental level of education"]!=("high school"or"some high school")]["average score"],color = "blue")
custom_lines = [Line2D([0], [0], color="red", lw=4),Line2D([0], [0], color="blue", lw=4)]
plt.legend(custom_lines,["high school level","higher education"])
plt.title("Average score by parental education")
```

We see that students with parents who have higher degree performed better in the test.
