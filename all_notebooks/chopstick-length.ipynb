## **Finding the best choice of each individual and store in 'best_lengths'.  
Make a dictinary with frequency of chopsticks' length.

```python
max_eff = np.zeros(31)
best_lengths = np.zeros(31)
for irow, row in df.iterrows():
    index = int(row['Individual'] - 1)
    if row['Food.Pinching.Efficiency'] > max_eff[index]:
        max_eff[index] = row['Food.Pinching.Efficiency']
        best_lengths[index] = int(row['Chopstick.Length'])
        
print(best_lengths)
Counter_lengths = dict(Counter(best_lengths))
print(Counter_lengths)
```

```python
x_val = list(Counter_lengths.keys())
y_val = list(Counter_lengths.values())
```

```python
sns.barplot(x = x_val, y = y_val)
```

### Chopsticks with length 240 had maximum FPE for majority of customers.

## **Finding the worst choice of each individual and store in 'worst_lengths'.  
Make a dictinary with frequency of chopsticks' length.

```python
min_eff = np.full(31, 200)
worst_lengths = np.zeros(31)
for irow, row in df.iterrows():
    index = int(row['Individual'] - 1)
    if row['Food.Pinching.Efficiency'] < min_eff[index]:
        min_eff[index] = row['Food.Pinching.Efficiency']
        worst_lengths[index] = int(row['Chopstick.Length'])
        
print(worst_lengths)
Counter_bad_lengths = dict(Counter(worst_lengths))
print(Counter_bad_lengths)
```

```python
x_bad = list(Counter_bad_lengths.keys())
y_bad = list(Counter_bad_lengths.values())
```

```python
sns.barplot(x = x_bad, y = y_bad)
```

### Chopsticks with length 180 gave least FPE for majority of customers.
