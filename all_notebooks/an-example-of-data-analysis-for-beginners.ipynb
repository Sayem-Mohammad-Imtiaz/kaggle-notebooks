```python
# opening the file world-happiness/2019.csv

data = pd.read_csv('/kaggle/input/world-happiness/2019.csv')

# reading the first 5 entries. If you want to see a specific number of entries you can write the numer into the bracets, for example data.head(10).

data.head()
```

```python
# The correlation between the attributes. 0 means no correlation. +1 (positive) and -1 (negative) mean there is a strong correlation.

data.corr()
```

As you can see from the results above, there are some high correlations between some attributes. Note that, overall rank has negative correlation between the other attributes because the most happy country's ranking is 1. The smaller number the ranking is, the happier the people in that country.  

* Overall rank - Score (-0,99), GDP per capita (-0,80), Social support (-0,76), Healthy life expectancy (-0,79) 
* Score - Overall rank (-0.99), GDP per capita (0.79), Social support (0.78), Healthy life expectancy (0.78)
* GDP per capita - Overall rank (-0.80), Score (0.79), Social support (0.75), Healthy life expectancy (0.86)
* Social support - Overall rank (-0.77), Score (0.78), GDP per capita (0.74)
* Healthy life expectancy -  Overall rank (-0.79), Score (0.78), GDP per capita (0.83), Social support (0,72)

Freedom to make life choices, Generosity and Perceptions of corruption aren't that much correlated with the other attributes.

```python
# To see the correlation between attributes with a correlation map
f,ax = plt.subplots(figsize=(10, 10))
sns.heatmap(data.corr(), annot=True, linewidths=.3, fmt= '.1f',ax=ax)
plt.show()
```

```python
# You can see the distribution of the Overall rank and GDP per capita by creating a scatter plot  
data.plot(kind = "scatter",x="Overall rank",y = "GDP per capita", color = "r")
plt.show()
```
