# Country-Wise Analysis

**SUICIDE RATE PER 100K PEOPLE**

```python
print("Country With Highest Suicide Rate Per 100K Persons")
df[df["suicideper100th"]==df["suicideper100th"].max()]["country"]
```

```python
print("Country With Lowest Suicide Rate Per 100K Persons")
df[df["suicideper100th"]==df["suicideper100th"].min()]["country"]
```

**ALCOHOL CONSUMPTION PER ADULT**

```python
print("Country With Highest Alcohol Consumption Per Capita")
df[df["alcconsumption"]==df["alcconsumption"].max()]["country"]
```

```python
print("Country With Lowest Alcohol Consumption Per Capita")
df[df["alcconsumption"]==df["alcconsumption"].min()]["country"]
```
