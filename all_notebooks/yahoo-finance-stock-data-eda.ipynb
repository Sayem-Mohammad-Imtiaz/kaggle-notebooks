```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Part 0
**Data reading and cleaning**

```python
data = pd.read_csv("../input/yahoo-finance-data/Yahoo_Finance.csv").dropna()
data.head()
```

```python
data.info()
```

```python
data = data[list(data.columns)[1:]]
data.info()
```

# Part 1
**General data preprocessing**

```python
year_data = data.copy()
month_data = data.copy()
day_data = data.copy()
```

```python
year_data["Year"] = 1
month_data["Month"] = 1
day_data["Day"] = 1
```

```python
for i in year_data.index:
    year_data["Year"][i] = float( year_data["Date"][i].split("-")[0] )
for i in month_data.index:
    month_data["Month"][i] = float( month_data["Date"][i].split("-")[1] )
for i in day_data.index:
    day_data["Day"][i] = float( day_data["Date"][i].split("-")[2] )
```

# Part 2
**Visualization**

```python
def plot_it(x, y, xlabel, ylabel):
    plt.figure(figsize=(10, 10))
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.plot(x, y)
    plt.show()
```

```python
for col in list(data.columns)[1:]:
    plot_it(data["Date"], data[col], "Date", col)
```
