## 2

```python
df['YearsExperience'].values.reshape(30,1)
```

```python
X=df['YearsExperience'].values.reshape(30,1)
```

```python
df['Salary'].values
```

```python
y=df['Salary'].values
```

## 3

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## 4

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

## This is the mostly perfect line

```python
X_test
```

```python
y_test
```

```python
reg.predict(X_test[0].reshape(1,1))
```

```python
reg.predict(X_test[1].reshape(1,1))
```

```python
reg.predict(X_test[5].reshape(1,1))
```

```python
reg.predict(np.array([[40]]))
```
