```python
cond =[(df.year>=1920) & (df.year<=1929), (df.year>=1930) & (df.year<=1939), (df.year>=1940) & (df.year<=1949), (df.year>=1950) & (df.year<=1959), (df.year>=1960) & (df.year<=1969), (df.year>=1970) & (df.year<=1979),(df.year>=1980) & (df.year<=1989), (df.year>=1990) & (df.year<=1999), (df.year>=2000) & (df.year<=2009), (df.year>=2010) & (df.year<=2019),(df.year>=2020)]
decade = ['1920','1930','1940','1950','1960','1970','1980','1990','2000','2010','2020']


df['decade'] = np.select(cond, decade)

df
```

# **Padronização dos dados**

Os algoritmos de clusterização são baseados na distância dos pontos intra e inter clusters. Por isso é necessário padronizar os dados para que todos fiquem na mesma escala.

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
fit=scaler.fit(df_cluster)
transform=scaler.transform(df_cluster)
print(transform)
```
