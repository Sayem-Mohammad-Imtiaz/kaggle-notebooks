```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')

font = {'size' : 20,
        'weight' : 'bold',
        'family' : 'sans-serif'}
plt.rc('font', **font)
```

```python
df = pd.read_csv('/kaggle/input/covid-vaccination-india-district-wise-data/vaccine_doses_statewise.csv')
```

```python
df.head()
df = df.set_index('State')
```

```python
df.head()
```

```python
vacc = df.T.reset_index()
vacc.head()
```

```python
vacc['index'] = pd.to_datetime(vacc['index'], dayfirst = True)
```

```python
vacc.head()
```

```python
vacc[['index', 'Punjab']].set_index('index').plot(figsize = (15, 8))
```

```python
vacc.tail()
```

```python
punjab_vacc = -(vacc['Punjab'].shift(1).fillna(0.0) - vacc['Punjab'])
punjab = vacc['Punjab']
punjab_vacc
```

```python
vacc['Punjab']
```

```python
df_punjab = pd.concat([vacc['index'], punjab, punjab_vacc], 1)

df_punjab.columns = ['Dates', 'Cumulative', 'Daily']
df_punjab.set_index('Dates')['Daily'].plot(figsize = (15, 8), color = 'maroon')
plt.vlines(['2021-05-01 00:00:00', '2021-06-21 00:00:00'], ymin = 0.0, ymax = 6*10**5)
```
