# Analysis on Crimes against Women in UP

```python
d1 = df[10].value_counts().to_dict()
d1
```

```python
d2 = df[11].value_counts().to_dict()
d2
```

```python
d3 = df[12].value_counts().to_dict()
d3
```

```python
for key in d1:
    if key in d2:
        d1[key] = d2[key] + d1[key]
    if key in d3:
        d1[key] = d3[key] + d1[key]
    else:
        pass
```

```python
d1
```

```python
l = list(d1.items())
l.sort()
d1 = dict(l)
d1
```

```python
d1.pop(0)
d1
```

```python
columns
```

```python
reas = ["Murder with Rape", "Dowry Deaths(Sec. 3048)","Suicide(sec 305/306)", "Kidnapping(All)", "Acid Attack(Sec. 326A IPC)", "Cruelty by Husband/in-laws((Sec.498 A IPC)", "Rape only(Sec. 376 or 511 IPC)", "Assault on Women with Intent to Outrage her Modesty (Sec. 354 IPC)", "Cyber Crimes against Women", "Protection of Children from Sexual Offences Act"]
reas
```

```python
d1.pop(11)
```

```python
fig = plt.figure(figsize = (20, 10))
plt.barh(reas, list(d1.values()))
plt.ylabel("Crime Against Women")
plt.xlabel("Frequency of Crime")
plt.title("Crime Against Women in UP")
```

**Assault on Women with Intent to Outrage her Modesty** and **Rape** are the major crimes that happen with women in UP.
