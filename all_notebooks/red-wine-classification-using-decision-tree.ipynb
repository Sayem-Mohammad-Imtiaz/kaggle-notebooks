```python
data = pd.read_csv("/kaggle/input/red-wine-quality-cortez-et-al-2009/winequality-red.csv")
```

```python
# creating X and y

X = data.drop(columns = 'quality')
y = data['quality']
```

```python
# splitting data into training and testing data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 355)
```

```python
# scaling our data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

## Let's try to increase Accuracy of our model by hyper parameter tuning.

```python
param = {
    'criterion': ['gini', 'entropy'],
    'max_depth': range(2, 32, 1),
    'min_samples_leaf': range(1, 10, 1),
    'min_samples_split': range(2, 10, 1),
    'splitter': ['best', 'random']
}
```

```python
from sklearn.model_selection import GridSearchCV

grid_search = GridSearchCV(estimator=clf, param_grid = param, cv = 5, n_jobs = -1)
grid_search.fit(X_train, y_train)
```

```python
# best parameters 

grid_search.best_params_
```

```python
# best score 

grid_search.best_score_
```

```python
clf = DecisionTreeClassifier(criterion = 'entropy', max_depth = 27, min_samples_leaf = 1, min_samples_split = 2, splitter = 'random')
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
