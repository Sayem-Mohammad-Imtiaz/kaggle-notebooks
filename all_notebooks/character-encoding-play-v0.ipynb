```python
import chardet
import ftfy
A='../input/character-encoding-examples/die_ISO-8859-1.txt'
B='../input/character-encoding-examples/file_guide.csv'
C='../input/character-encoding-examples/harpers_ASCII.txt'
D='../input/character-encoding-examples/olaf_Windows-1251.txt'
E='../input/character-encoding-examples/portugal_ISO-8859-1.txt'
F='../input/character-encoding-examples/shisei_UTF-8.txt'
G='../input/character-encoding-examples/yan_BIG-5.txt'

files=[A,B,C,D,E,F,G]
for file in files:
    filename=file[file.index('s')+2:file.index('s')+6]
    print(filename)
    with open(file, 'rb') as rawdata:
        result = chardet.detect(rawdata.read(100000))
        print(result)
        print(result['encoding'])
    df = pd.read_csv(file,encoding=result['encoding'],sep='|')
    df.loc[0]=df.loc[0].apply(lambda s: ftfy.fix_text(s))
    #ftfy.fix_text(s)
    print(df.head())
    df.to_csv(filename+'.csv')
```
