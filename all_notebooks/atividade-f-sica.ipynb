**Introdução**
    O sedentarismo se define pela ausência de práticas de atividade física ao longo do dia e ocasiona uma diminuição do gasto calórico. Esse comportamento é intendificado pelo estilo de vida inapropriado como por exemplo, muito tempo na posição sentada ou deitada e ainda se alimentando de comidas nada saudáveis durante esse tempo de imobilismo. 
    Atualmente um terço da população mundial adulta é inativa fisicamente e isso gera 5 milhoes de mortes por ano (The LAncet, 2012). Além de contribuir para diversas doenças crônicas, o sedentarismo também influencia no humor, na qualidade do sono e no peso corporal.
    Dessa forma, os objetivos da escolha dessa base de dados foram:
* Verificar se existem associações entre atividade física (pela contagem de passos),gasto calórico,peso corporal horas de sono e na sensação de sentir-se ativo e/ou inativo.
* Comparar o gasto calórico entre as categorias de humor e de auto percepção de atividade (ativo e inativo)
* Comparar as horas de sono entre as categorias de humor e de auto percepção de atividade (ativo e inativo)
* Comparar o peso corporal entre as categorais de auto percepção de atividade (ativo e inativo)
    
    **base de dados **
   A base de dados possui 96 observações, 7 colunas. Tem como variáveis quantitativas "número de passos"(step_count), "gasto calórico"(calories_burned), "horas de sono" (hours_of_sleep e "peso corporal" (weight_kg). E como variáveis qualitativas "datas" (date), "humor" (mood), auto percepção de atividade "ativo ou inativo" (bool_of_active). Sendo que para a variável "humor" foi atribuído o valor "300" para significar "Feliz", o valor "200" para "Neutro" e "100" para "triste" e para a variável "auto percepção de atividade" foi atribuido o valor de "500" para "ativo" e "0" para inativo.
   
   **Metodologia**
*    Será expresso em médias e desvios-padrão variáveis quantitativas
*    Será exposto tabelas de contigência de variáveis categóricas.
*    Será apresentado uma matriz de correlação entre as variáveis
*    Gráficos de barras e violinos para demonstrar a distribuição de variáveis quantitativas por categorias
*    Gráfico de dispersão para análise de possível relação linear entre duas variáveis

**RESSALVAS**

```python
#GRÁFICO DE CORRELAÇÃO JUNTAMENTE COM A DISTRIBUIÇÃO - STEP_COUNT X CALORIES_BURNED
Correlacao= df[['step_count','calories_burned','hours_of_sleep','weight_kg', 'mood', 'bool_of_active']]
Correlacao
sns.jointplot(data=Correlacao,y='calories_burned',x='step_count',kind='reg', color='g')
```

# **RESSALVAS**
Apesar da forte relação entre a contagem de passos com as calorias gastas (r = 0.98) observa-se uma distribuição assimétrica de ambas variáveis.

```python
#GRÁFICO DE CORRELAÇÃO JUNTAMENTE COM A DISTRIBUIÇÃO - MOOD X WEIGTH_KG
sns.jointplot(data=Correlacao,y='weight_kg',x='mood',kind='reg', color='b')
```

**RESSALVAS**
Apesar de existir também uma correlação negativa e moderada entre "humor" e "peso corporal" ( r= -0.45) é necessário observar a curva de distribuição da variável categórica de humor "mood", por ser uma variável categórica, sugere-se realizar uma inferencia não paramétrica como a correlação de Spearman para verificar se essa correlação é significante.


```python
#verificando a estatística descritiva_média da contagem de passos por categoria de humor (100 = triste, 200 = neutro e 300 = feliz)
df.step_count.groupby(df.mood).describe()
```

```python
#gráfico de barras que demonstra a distribuição da contagem de passos por categoria de humor (100 = triste, 200 = neutro e 300 = feliz)
sns.barplot(x = "mood", y="step_count", data = df)
plt.title('Gráfico 1. Comparação da contagem de passos por percepção do humor')
plt.legend(['100 = Triste', '200 = Neutro', '300 = Feliz'], loc='upper left', prop={'size': 7})
plt.figure(figsize=(6,5))
#gráfico violino que demonstra a distribuição da contagem de passos por categoria de humor (100 = triste, 200 = neutro e 300 = feliz)
sns.violinplot(df['mood'], df['step_count'])
plt.title('Gráfico 2. Comparação da contagem de passos por percepção do humor')
plt.legend(['100 = Triste', '200 = Neutro', '300 = Feliz'], loc='upper left', prop={'size': 7})
```

```python
#verificando a estatística descritiva_média da contagem de passos por auto percepção de atividade (0 = inativo e 500 = ativo)
df.step_count.groupby(df.bool_of_active).describe()
```

```python
#criando uma tabela de contigencia entre sentir-se ativo ou inativo (bool_of_activity) x humor (mood) em percentual
pd.crosstab([df. bool_of_active], [df.mood], normalize='index', margins=True)
```

**CONCLUSÃO**
    Mediante a visualização de dados pode se observar que existe um correlação forte e positiva entre contagem de passos e calorias gastas, indicando que quanto mais o indivíduo se movimento melhor será o nível de atividade física. Bem como observou-se uma associação moderada e negativa entre peso corporal (kg) com o humor, isso indica que quanto mais massa corporal o indivíduo possui mais tendência a tristeza.
    Foi observado que aos se comparar o nível de atividade física pela contagem de passos, pessoas auto-denominada "felizes" demonstravam maior valor quando comparada com as categorias de "triste" e de "neutro).Isso demonstra o quando se movimentar o sistema nervoso  aumenta a produção de neurotransmissores do bem estar e da felicidade como dopamina, serotonina, ocitocina e endorfina.Adicionalmente, pessoas "felizes" gastam mais energia (kilocalorias) que pessoas auto-denominadas "neutras" ou "tristes". Em vista disso, as pessoas "felizes" demonstrar dormir mais horas de sono quando comparada com as demais categorias, no entanto, vale ressaltar que pessoas consideradas triste também dormem mais que as pessoas neutras. Isso reforça um dos sintomas de pessoas com quadro clínico de depressão terem preferencia por dormir mais tempo para aliviar o sofrimento.
    A auto percepção de atividade (ativou / inativo) também demonstra diferenças no caso do nível de atividade fisica (pela contagem de passos), pessoas "ativas" tendem a caminhar mais passos, gastar mais calorias e dormir mais horas quando comparados com as pessoas auto-denominadas "inativas". O estudo de Jessica Selinger (Current Biology, 2015) relata sobre a "lei do menor esforço" e demonstra que o corpo se ajusta aos menores esforços, pois o cérebro "boicota" os esforços para poupar gastos energéticos. Por isso, exigir fisicamente do corpo melhora a saúde e aumenta a disposição.
    Ao verificar a associação entre auto-percepção de atividade com humor, 46% das pessoas consideradas "tristes" também são "inativas" e o contrário ocorre, pois 57% das pessoas "felizes" são consideradas "ativas. Mediante ao exposto, conclui-se que a prática de atividade física regular é intervenção protagonista para uma melhor saúde física e mental.
