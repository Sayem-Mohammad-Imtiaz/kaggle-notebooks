# Enhancing the data set

```python
df2 = pd.read_csv('/kaggle/input/percapitaincome/csvData.csv')
```

```python
df2.head()
```

```python
df2.columns = ['country', 'incomeperperson','medianPerCapitaIncome','medianAnnualIncome','pop2021']
df['incomeperperson'] = df['incomeperperson'].combine_first(df2['country'].map(df.set_index('country')['incomeperperson']))
```

```python
df[df['incomeperperson'].isnull()]
```

```python
df.isnull().sum()
```

```python
df3=pd.read_csv('/kaggle/input/country-continent-mapping/continents2.csv')
```

```python
df3.head()
```

```python
df3.rename(columns={'name':'country'}, inplace=True)
```

```python
df3.head()
```

```python
left_merged = pd.merge(df, df3[['country','region','sub-region']],
                        how="left", on=["country"])
```

```python
left_merged.head()
```

```python
left_merged.isnull().sum()
```

```python
left_merged[left_merged['region'].isnull()]
```

```python
left_merged['region'].unique()
```

```python
data1 = left_merged['region'].value_counts()
```

```python
data1.plot.pie(autopct="%.1f%%");
```

```python
left_merged.head()
```

```python
grouped_data = left_merged.groupby(['region','sub-region'])
grouped_data.agg([np.mean,np.max,np.min])
```
