```python
movies=pd.read_csv('/kaggle/input/disney-movies-19372016-total-gross/disney_movies_total_gross.csv')
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

 Genre and mpaa_rating contains null values

```python
movies.fillna(np.NaN,inplace=True)
```

```python
# converting to datetime
movies['release_date'] = pd.to_datetime(movies['release_date'])
```

```python
movies['year'] = movies['release_date'].dt.date.astype(str).str.split('-', expand=True)[0]
```

```python
movies.head()
```

```python
# top 5 movies with best total_gross
movies.sort_values('total_gross',ascending=False).head(5)
```

```python
# top 5 movies with best inflation_adjusted_gross
movies.sort_values('inflation_adjusted_gross',ascending=False).head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.style.use('dark_background')
```

```python
sns.countplot(y='genre',data=movies)
```

count of comedy movies are more in numbers

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sns.histplot(movies['year'],kde=True)
```

This plot shows that most movies were released in year 1995

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sns.histplot(x='year',hue='genre',data=movies)
```

```python
sns.barplot(y='genre',x='total_gross',data=movies)
```

Adventurous movies have max total gross

```python
sns.barplot(y='genre',x='inflation_adjusted_gross',data=movies)
```

musical category has most inflation adjusted gross

```python
plt.figure(figsize=(12,6))
sns.boxplot(y='genre',x='inflation_adjusted_gross',data=movies)
```

musical category has many outliers in adjusted gross

```python
sns.countplot(x='mpaa_rating',data=movies)
```

number of PG ratings are more in number

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sns.lineplot(x='year', y='inflation_adjusted_gross', data=movies)
```

```python
plt.figure(figsize=(12,8))
plt.xticks(rotation=90)
sns.lineplot(x='year', y='total_gross', data=movies)
```

total gross is increasing gradually over years but it is not the same case with adjusted gross

```python
plt.figure(figsize=(8,4))
sns.countplot(x='mpaa_rating',hue='genre',data=movies)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.lmplot(x='total_gross',y='inflation_adjusted_gross',data=movies)
```
