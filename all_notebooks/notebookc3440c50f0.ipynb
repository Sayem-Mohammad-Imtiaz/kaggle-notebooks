```python
fig, ax1 = plt.subplots(1,1,figsize=(16,7), dpi= 80)
ax1.plot(data['SYMBOL'],data['OPEN'],color='tab:red')
ax1.set_xlabel('SYMBOL', fontsize=20)
ax1.set_ylabel('OPEN', color='tab:red', fontsize=20)
ax1.tick_params(axis='x', rotation=90 )
ax2 = ax1.twinx()
ax2.plot(data['SYMBOL'],data['CLOSE'],color='tab:green')
ax2.set_xlabel('SYMBOL', fontsize=20)
ax2.set_ylabel('CLOSE', color='tab:green', fontsize=20)
ax2.tick_params(axis='x', rotation=90 )

plt.show()
```

```python
fig, ax1 = plt.subplots(1,1,figsize=(16,7), dpi= 80)
ax1.plot(data['SYMBOL'],data['HIGH'],color='tab:red')
ax1.set_xlabel('SYMBOL', fontsize=20)
ax1.set_ylabel('HIGH', color='tab:red', fontsize=20)
ax1.tick_params(axis='x', rotation=90 )
ax2 = ax1.twinx()
ax2.plot(data['SYMBOL'],data['LOW'],color='tab:green')
ax2.set_xlabel('SYMBOL', fontsize=20)
ax2.set_ylabel('LOW', color='tab:green', fontsize=20)
ax2.tick_params(axis='x', rotation=90 )

plt.show()
```

```python
fig,axis=plt.subplots(2,3,figsize=(20,12))
sns.lineplot(data=data,x="DATE",y="PREV CLOSE",ax=axis[0,0],color="red")
sns.lineplot(data=data,x="DATE",y="OPEN",ax=axis[0,1],color="black")
sns.lineplot(data=data,x="DATE",y="HIGH",ax=axis[0,2],color="blue")
sns.lineplot(data=data,x="DATE",y="LOW",ax=axis[1,0],color="green")
sns.lineplot(data=data,x="DATE",y="LAST",ax=axis[1,1],color="yellow")
sns.lineplot(data=data,x="DATE",y="CLOSE",ax=axis[1,2],color="brown")
```
