**Why normalization?**
Because when we normalize the data we ensure that every data feature is between the values we want without destroying the differences between the points.By doing that we avoid some data features' influence more because they are on a different scale.

**Why hot encoding instead of indexing?**
Because if you index instead of one hot encoding it becomes hard or impossible to make mathematical inferences.Since you will multiply a value in a column with an weight if you use indexing you create a non-existent relation.

```python
def min_max_scaling(column):
    column1=column.copy()
    column1= (column-column.min())/(column.max()-column.min())
    return column1
#i will use these four later
cti_max=data["city_development_index"].max()
cti_min=data["city_development_index"].min()
th_max=data["training_hours"].max()
th_min=data["training_hours"].min()
def returning_to_past(column,maxd,mind):
    column1=column.copy()
    column1=np.add(np.multiply((maxd-mind),column1),mind)
    return column1
data["city_development_index"]=min_max_scaling(data["city_development_index"])
data["training_hours"]=min_max_scaling(data["training_hours"])

data
```
