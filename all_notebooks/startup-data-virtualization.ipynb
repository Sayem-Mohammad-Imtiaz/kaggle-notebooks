# Reading data

```python
data = pd.read_csv("../input/startup-investments-crunchbase/investments_VC.csv", encoding = 'unicode_escape')
data.rename(columns = {' market ': 'market', ' funding_total_usd ': 'funding_total_usd'}, inplace = True)
```

# Taking a quick look at the structure of data


```python
data.head()
```

```python
data.describe()
```

### Data cleaning

```python
data.drop(data[data['funding_total_usd'] == ' -   '].index,inplace=True)
data['funding_total_usd'] = data['funding_total_usd'].str.replace(",","")
data.dropna( axis=0,subset=['funding_total_usd'],inplace = True) and data.dropna( axis=0,subset=['market'],inplace = True)
data.drop('homepage_url', inplace = True, axis = 1)
data['funding_total_usd']=data['funding_total_usd'].astype('int')
marketdf = data[['market','name','funding_total_usd','status']]
```

# Virtualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
from plotly.offline import iplot, init_notebook_mode
from wordcloud import WordCloud
```

```python
plt.figure(figsize=(28,8))
sns.countplot(x='market',data=marketdf, order=marketdf['market'].value_counts().iloc[:15].index)
plt.xlabel('Markets')
plt.ylabel('Counts')
plt.title('Amount of startup by Markets')
```

### At the graph below we see the top 15 industries in ecosystem. As you can see most popular 2 one is "Software" and "Biotechnology", however It is too early to say the best options are these industries or success rate is the highest.

```python
mark=marketdf['market']
mark.dropna(inplace=True)
wordcloud = WordCloud(background_color = 'white',width = 5000,height = 3000).generate(" ".join(mark))
plt.figure(figsize=(20,10))
plt.imshow(wordcloud,interpolation='bilinear') 
plt.axis('off') 
plt.show()
```

```python
plt.figure(figsize=(25,8))
sns.barplot(x='market',y= 'funding_total_usd',data = marketdf,order=marketdf['market'].value_counts().iloc[:15].index, estimator=median)
plt.xlabel('Markets')
plt.ylabel('Medians of fundings')
plt.title('Median of fundings by markets')
```
