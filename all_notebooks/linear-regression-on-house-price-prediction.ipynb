# **TRAINING A LINEAR REGRESSION MODEL**

```python
X = Housing_data[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
               'Avg. Area Number of Bedrooms', 'Area Population']]
Y =Housing_data['Price']
```

# **TRAIN AND TEST SPLIT**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

# **EVALUATION AND CALCULATION OF R2 MSE RMSE MAE FROM ACTUAL AND PREDICTED VALUES**

```python
from sklearn import metrics
from sklearn.model_selection import cross_val_score

def cross_val(model):
    pred = cross_val_score(model, X, Y, cv=10)
    return pred.mean()

def print_evaluate(true, predicted):  
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R2 Square', r2_square)
    print('__________________________________')
    
def evaluate(true, predicted):
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    return mae, mse, rmse, r2_square
```

# **LINEAR REGRESSION**

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression(normalize=True)
lin_reg.fit(X_train,Y_train)
```

# **MODEL EVALUATION PROCESS**

```python
# print the intercept
print(lin_reg.intercept_)
```

```python
coeff_df = pd.DataFrame(lin_reg.coef_, X.columns, columns=['Coefficient'])
coeff_df 
```

# **PREDICTIONS FROM MY MODEL**

```python
Y_pred = lin_reg.predict(X_test)
plt.scatter(Y_test,Y_pred)
```

# **RESIDUAL HISTOGRAM**

```python
sns.distplot((Y_test -Y_pred), bins=50);
```

# **REGRESSION EVALUATION METRICS**

```python
Y_Test_pred = lin_reg.predict(X_test)
Y_Train_pred = lin_reg.predict(X_train)

print('Test set evaluation:\n_____________________________________')
print_evaluate(Y_test, Y_Test_pred)
print('Train set evaluation:\n_____________________________________')
print_evaluate(Y_train,Y_Train_pred)
```

# **RESULTS FROM MODEL CREATION**

```python
res_df=pd.DataFrame(data=[["Linear Regression", *evaluate(Y_test,Y_Test_pred),cross_val(LinearRegression())]] ,columns=['Model','MAE','MSE','RMSE','R2 square','Cross validation'])
res_df                 
                   
```
