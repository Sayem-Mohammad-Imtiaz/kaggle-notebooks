```python
import nltk
import re
import matplotlib.pyplot as plt
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
fake=pd.read_csv('../input/fake-and-real-news-dataset/Fake.csv')
true=pd.read_csv('../input/fake-and-real-news-dataset/True.csv')
```

```python
true.head()
```

```python
fake['label']=0
true['label']=1
```

```python
fake_df=fake.drop(['title','subject','date'],1)
true_df=true.drop(['title','subject','date'],1)
```

```python
data=pd.concat([fake_df.iloc[:1000,:],true_df.iloc[:1000,:]],ignore_index=True)
```

```python
data.head()
```

```python
x=data.drop('label',1)
y=data['label']
```

```python
ps=PorterStemmer()
corpus=[]
for i in range(len(x)):
    review=re.sub('[^a-zA-Z]'," ",x['text'][i])
    review=review.lower()
    review=review.split(" ")
    review=[ps.stem(word) for word in review if word not in stopwords.words('english')]
    review=" ".join(review)
    corpus.append(review)
```

```python
vector=TfidfVectorizer(max_features=3000)
x=vector.fit_transform(corpus).toarray()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix,plot_confusion_matrix,accuracy_score
```

```python
clf=MultinomialNB()
clf=clf.fit(x_train,y_train)
pred=clf.predict(x_test)
```

```python
acc=accuracy_score(y_test,pred)
print('Accuracy for the model is ',acc*100)
```

```python
plot_confusion_matrix(clf,x_test,y_test)
plt.show()
```
