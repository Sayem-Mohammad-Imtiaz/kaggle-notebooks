```python
dataset = pd.read_csv("../input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv")
```

```python
dataset
```

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
data = []
for i in range(0,len(dataset)):
    review = dataset["review"][i]
    review = re.sub('[^a-zA-Z]',' ',review)
    
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in set(stopwords.words("english"))]
    review = ' '.join(review)
    
    data.append(review)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=2000)
x = cv.fit_transform(data)
x = x.toarray()

y = dataset.iloc[:,1:2].values
for i in range(0,len(y)):
    if y[i] == "positive":
        y[i] = 1
    else:
        y[i] = 0
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(units=1565,kernel_initializer="random_uniform",activation="relu"))
model.add(Dense(units=3000,kernel_initializer="random_uniform",activation="relu"))
model.add(Dense(units=3000,kernel_initializer="random_uniform",activation="relu"))
model.add(Dense(units=1,kernel_initializer="random_uniform",activation="sigmoid"))
model.compile(optimizer="rmsprop",loss="binary_crossentropy",metrics=["accuracy"])
history=model.fit(x_train.astype(np.float),y_train.astype(np.float),epochs=10)
```

```python
print(history.history["accuracy"])
```

```python
pred = model.predict(x_test)
pred=pred>0.5
pred
```
