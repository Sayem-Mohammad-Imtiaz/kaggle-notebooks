# Importing Dataset


```python
data = pd.read_csv("/kaggle/input/breast-cancer-csv/breastCancer.csv")
```

# Data Pre-processing

```python
data.head(5)
```

```python
data.dtypes
```

```python
data["bare_nucleoli"].value_counts()
data = data.replace("?",np.nan)
data = data.fillna(data.median())
data["bare_nucleoli"]= data["bare_nucleoli"].astype("int64")
```

# Building Model (KNN and SVM)

> Importing train_test_split module

```python
from sklearn.model_selection import train_test_split
```

```python
X = data.drop("class",axis=1)
Y = data["class"]
```

```python
train_x,test_x,train_y,test_y = train_test_split(X,Y,test_size= 0.33, random_state=42)
```

# KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier(n_neighbors=5)
KNN.fit(train_x,train_y)
pred1 = KNN.predict(test_x)
```

```python
Accuracy_knn = KNN.score(test_x,test_y)*100
print ("accuracy of the model is {0:.2g}%".format(Accuracy_knn))
```

# SVM classifier

```python
from sklearn.svm import SVC
```

```python
SVM = SVC()
SVM.fit(train_x,train_y)
pred2 = SVM.predict(test_x)
```

```python
SVM.score(test_x,test_y)
```

# Classification metric

```python
from sklearn.metrics import classification_report
```

```python
print ("KNN classifier")
print (classification_report(test_y,pred1))
```

```python
print ("SVM classifier")
print (classification_report(test_y,pred2))
```
