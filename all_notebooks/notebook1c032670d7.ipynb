Data cleaning

```python
#count number of null(empty) values
df.isna().sum()
```

```python
#drop the column with nan values
df.dropna(axis=1,inplace=True)
df.shape
```

```python
# Get the total numbr of M & B cells in diagnosis
df['diagnosis'].value_counts()
```

Label Encoding

```python
# get datype in each colums
df.dtypes
```

```python
# encode the daignosis values
from sklearn.preprocessing import  LabelEncoder
labelEncoder= LabelEncoder()
df.iloc[:,1]=labelEncoder.fit_transform(df.iloc[:,1].values)
df
```

```python
# split the data into dependent and independen dataset.
x=df.iloc[:,2:].values
y=df.iloc[:,1].values
```

```python
# splitting the data into traning and testing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size=0.25)
```

```python
# data scaling 
```

Feature scaling


```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler();
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

Building logistic model

```python
from sklearn.linear_model import LogisticRegression
classifier= LogisticRegression()
classifier.fit(x_train, y_train)
```

```python
##make use of trained model to the make predictions on test data
prediction= classifier.predict(x_test);
```

performance evaluations

```python
#plot confussion matrix
from sklearn.metrics import  confusion_matrix
cm=confusion_matrix(y_test, prediction)
print(cm)
```

@ get accuracy model

```python
# get accuracy model

import seaborn as sns
sns.heatmap(cm, annot=True)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, prediction))
```
