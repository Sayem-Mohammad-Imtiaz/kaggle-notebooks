```python
import wittgenstein as lw
```

```python
df_train, df_test = train_test_split(df2, test_size=.20, random_state=42)
df_test_x = df_test.drop("% Silica Concentrate Bins", axis = 1)
df_test_y = df_test["% Silica Concentrate Bins"]
```

```python
# Ripper
clf = lw.RIPPER(k = 2, prune_size = 0.33, dl_allowance = 64, n_discretize_bins = None, random_state=42,
               max_rules = None, max_rule_conds = None, max_total_conds = None)
```

```python
clf.fit(df_train, class_feat="% Silica Concentrate Bins", pos_class="low")
```

```python
# Getting precision score
precision = clf.score(df_test_x, df_test_y, precision_score)
# Getting recall score
recall = clf.score(df_test_x, df_test_y, recall_score)
# Getting the number of rules
cond_count = clf.ruleset_.count_conds()
# print results
print(f'precision: {precision} recall: {recall} conds: {cond_count}')
```

```python
clf.ruleset_.out_pretty()
```

```python
df2_results = clf.predict(df2.drop("% Silica Concentrate Bins", axis = 1))
```

```python
df2["prediction"] = pd.Series(df2_results)
```
