```python
import geopandas as gpd
```

```python
import pandas as pd
```

```python
import folium
from folium import Choropleth
```

```python
def embed_map(m, file_name):
    from IPython.display import IFrame
    m.save(file_name)
    return IFrame(file_name, width='100%', height='500px')
```

```python
CA_high_earners = pd.read_csv("../input/geospatial-learn-course-data/CA_county_high_earners.csv", index_col='GEOID')
CA_high_earners.head()
```

```python
CA_pop = pd.read_csv("../input/geospatial-learn-course-data/CA_county_population.csv", index_col='GEOID')
CA_pop.head()
```

```python
CA_stats = CA_high_earners.join(CA_pop)
CA_stats.head()
```

```python
CA_stats["rate_of_rich"] = CA_stats["high_earners"] / CA_stats["population"]
CA_stats.head()
```

```python
CA_counties = gpd.read_file("../input/geospatial-learn-course-data/CA_county_boundaries/CA_county_boundaries/CA_county_boundaries.shp")
CA_counties.head()
```

```python
CA_stats2 = pd.merge(CA_counties, CA_stats, on='GEOID')
CA_stats2.head()
```

```python
# key_on = "feature.id" requires a string-type column. Here we use "name" for it.
```

```python
name_geometry = CA_stats2.drop(columns=['GEOID', 'area_sqkm', 'high_earners', 'population', 'rate_of_rich']).set_index('name')
name_geometry.head()
```

```python
name_rate_of_rich = CA_stats2.drop(columns=['GEOID', 'area_sqkm', 'high_earners', 'population', 'geometry']).set_index('name')
name_rate_of_rich.head()
```

```python
# name_geometry.__geo_interface__
```

```python
name_rate_of_rich
```

```python
type(name_rate_of_rich)
```

# data in Chropleth should be pd.Series.

```python
# Converting to pd.Series via Python lists.

rate_of_rich_Series = pd.Series(name_rate_of_rich['rate_of_rich'].values.tolist(), index=name_rate_of_rich.index.tolist())
type(rate_of_rich_Series)
```

```python
rate_of_rich_Series
```

```python
m_9 = folium.Map(location=[36.5,-120], tiles='cartodbpositron', zoom_start=5)
Choropleth(geo_data=name_geometry.__geo_interface__, 
           data=rate_of_rich_Series, 
           key_on="feature.id", 
           fill_color='YlGnBu', 
           legend_name='Rate of high earners'
          ).add_to(m_9)
embed_map(m_9, 'q_9.html')
```
