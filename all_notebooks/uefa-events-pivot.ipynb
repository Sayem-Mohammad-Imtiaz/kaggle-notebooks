```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

df = pd.read_csv('../input/uefa-2021-events/uefa_events.csv')

# we need to count the events and work with them
df['event_type'].value_counts().to_frame()
df['statics'] = df['event_type']

# dividing data into (home and away), so you can know the statistcs of each team when in Homegroun or in Away
df_home = df.query('event_team == "home"')
df_home = df[['team_home', 'event_type', 'statics']]
df_home = df_home.groupby(['team_home', 'event_type'], as_index=False)['statics'].count()
df_home.rename(columns=({'team_home' : 'Team Name'}), inplace=True)
df_home

##----------------------------------------
df_away = df.query('event_team == "away"')
df_away = df[['team_away', 'event_type', 'statics']]
df_away = df_away.groupby(['team_away', 'event_type'], as_index=False)['statics'].count()
df_away.rename(columns = ({'team_away' : 'Team Name'}),inplace=True)
df_away

###-----------------------------------------

# now you can add them togather to be able to extract the statistics for each team in total (home and away)

df_all = pd.concat([df_home, df_away]).groupby(['Team Name', 'event_type'], as_index=False).sum()
df_all = pd.DataFrame(df_all)
df_pivot = df_all.pivot(index = 'Team Name', columns = 'event_type')
df_pivot.replace(np.nan, '', inplace=True)
print(df_pivot.head())

# From here you can do your exploratory and conclusion as you want :).
```
