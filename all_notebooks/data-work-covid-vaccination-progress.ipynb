```python
covac = pd.read_csv('/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv')
```

```python
covac.info()
```

```python
covac['date'] = pd.to_datetime(covac['date'])
```

```python
covac['date'].iloc[0].day
```

```python
covac['date_month'] = covac['date'].apply(lambda x:x.month)
covac['date_dayofweek'] = covac['date'].apply(lambda x:x.dayofweek)
covac['date_day'] = covac['date'].apply(lambda x:x.day)
covac['date_year'] = covac['date'].apply(lambda x:x.year)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
# mmap = {1.0: 'Jan', 2.0: 'Feb', 3.0: 'Mar', 4.0: 'Apr', 5.0: 'May',
#             6.0: 'Jun', 7.0: 'Jul', 8.0: 'Aug', 9.0: 'Sep', 10.0: 'Oct', 11.0: 'Nov', 12.0: 'Dec'}
```

```python
covac['date_dayofweek'] = covac['date_dayofweek'].map(dmap)
```

```python
covac.head(2)
```

```python
covac_2021 = covac[covac['date_year'] == 2021]
day_month = covac_2021.groupby(['date_day','date_month']).sum()['daily_vaccinations'].unstack().fillna(0)
day_month
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(day_month,annot=True, cmap="crest", fmt='g')
```
