```python
data = pd.read_csv('/kaggle/input/spotify-dataset-19212020-160k-tracks/data.csv')

print(data)
```

```python
def search_artist(artist_name):
    re_expression = '(?<=Eminem.)'
    artist_name = str(artist_name)
    artist_name = artist_name.lower()
    is_the_artist = re.search(re_expression, artist_name)
    if is_the_artist:
        return True
    else:
        return False
```

```python
def search_year(release_date):
    re_expression = '(?<=1921.)'
    release_date = str(release_date)
    release_date = release_date.lower()
    is_the_year = re.search(re_expression, release_date)
    if is_the_year:
        return True
    else:
        return False
```

```python
artist_found = data[data['artists'].apply(search_artist)==True]
print(data['artists'])
#print(artist_found)
```
