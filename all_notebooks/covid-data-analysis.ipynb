```python
india_data.drop(['Sno', 'Time','ConfirmedIndianNational', 'ConfirmedForeignNational'],inplace = True, axis = 1)
```

```python
india_data = india_data.rename(columns={'State/UnionTerritory':'States','Cured':'Recovered'})
india_data.head()
```

```python
india_data['Active'] = india_data['Confirmed']-india_data['Recovered']-india_data['Deaths']
india_data.head()
```

```python
def confirmed_cases(state):
    india_data_new = india_data[india_data['States']==state]
    return india_data_new
```

```python
columns = ['Recovered', 'Deaths', 'Confirmed', 'Active']
for i in columns:
    fig = px.treemap(india_data,values=i,path=['States'],template="plotly_dark",title="<b>TreeMap representation of different States w.r.t. their {}</b>".format(i))
    fig.show()
```

```python
def plot_fig(x,y):
    fig=px.line(india_data,x=x,y=y,title="Covid cases " + str(x) + ' vs ' + str(y),template="plotly_dark")
    return fig.show()
```

```python
plot_fig('Date','Recovered')
```

```python
plot_fig('Date','Deaths')
```

```python
plot_fig('Date','Confirmed')
```

```python
plot_fig('Date','Active')
```
