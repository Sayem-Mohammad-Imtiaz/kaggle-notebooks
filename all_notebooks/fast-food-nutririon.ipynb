We also got ' -   ' for Saturated Fat and Trans Fat, it is 0 when I cross checked the data. Lets replace these and read the data into a new df - food1

```python
food1 = food.copy()
```

```python
food1.head()
```

```python
food1[["Total Fat (g)", "Saturated Fat (g)", "Trans Fat (g)"]].astype(str) 
```

```python
food1["Total Fat (g)"] = food1["Total Fat (g)"].str.replace(',','.')
```

```python
food1["Saturated Fat (g)"] = food1["Saturated Fat (g)"].str.replace(',','.')
```

```python
food1["Trans Fat (g)"] = food1["Trans Fat (g)"].str.replace(',','.')
```

```python
food1["Total Fat (g)"] = food1["Total Fat (g)"].str.replace(' -   ','0')
```

```python
food1["Saturated Fat (g)"] = food1["Saturated Fat (g)"].str.replace(' -   ','0')
```

```python
food1["Trans Fat (g)"] = food1["Trans Fat (g)"].str.replace(' -   ','0')
```

```python
print(food1["Trans Fat (g)"].unique());
print(food1["Saturated Fat (g)"].unique());
print(food1["Trans Fat (g)"].unique());
```

```python
food1[["Total Fat (g)", "Saturated Fat (g)", "Trans Fat (g)"]] = food1[["Total Fat (g)", "Saturated Fat (g)", "Trans Fat (g)"]].astype(float)
```

```python
food1.info()
```

Serving size should be a numerical column too. Lets get to it.

```python
food1["Serving Size (g)"]= food1["Serving Size (g)"].fillna(0)
```

```python
food1["Serving Size (g)"].unique()
```

```python
food1["Serving Size (g)"] = food1["Serving Size (g)"].astype(str)
```

```python
food1["Serving Size (g)"] = food1["Serving Size (g)"].str.replace(",",".")
```

```python
food1["Serving Size (g)"] = food1["Serving Size (g)"].astype(float)
```
