```python
# pacakages required 
import numpy as np 
import pandas as pd 
import os 


#importing the required csv files
for dirname, _, filenames in os.walk('/kaggle/input'):
    if filenames != []:
        cust_details = pd.read_csv(os.path.join(dirname, filenames[0]))
        transac_details = pd.read_csv(os.path.join(dirname, filenames[1]))

#one-hot encoding transaction details
df1 = pd.get_dummies(transac_details['paymentMethodType'])
df2 = pd.get_dummies(transac_details['paymentMethodProvider'])
df3 = pd.get_dummies(transac_details['orderState'])
df4 = pd.get_dummies(transac_details['transactionFailed'])
df5 = pd.get_dummies(transac_details['paymentMethodRegistrationFailure'])

#merge and drop unecessary columns in transaction details
transac=transac_details[['customerEmail','transactionAmount']]
transac_details = pd.concat([transac,df1,df2,df3,df4,df5], axis=1)
transac_details = transac_details.groupby(['customerEmail']).sum()

# dropping duplicates in customer details 
cust_details = cust_details.drop_duplicates(subset='customerEmail', keep='first')

#merging customer and transaction details 
data=pd.merge(cust_details,transac_details,on='customerEmail')


#dropping unecesarry rows 
data=data.drop(columns=['Unnamed: 0','customerEmail', 'customerPhone','customerDevice','customerIPAddress','customerBillingAddress'])
data.columns = ['No_Transactions','No_Orders','No_Payments','Fraud','transactionAmount','apple pay','bitcoin','card','paypal','American Express','Diners Club / Carte Blanche','Discover','JCB 15 digit','JCB 16 digit','Maestro','Mastercard','VISA 13 digit','VISA 16 digit','voyager','failed','fulfilled','pending','transaction_success','transaction_failed','reg_success','reg_failed']
Y=data['Fraud']
data=data.drop(columns=['Fraud'])
X=data
#data=pd.concat([data,f],axis=1)
#data.to_csv("data.csv")
print(data.dtypes)
print(data.head())


from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size=0.25, random_state=0)
```

1. Logistic Regression
2. Naive Bayes
3. Desicion Tree
4. SVM
5. Random Forest
6. KNN

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(max_iter=1000)
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
score = lr.score(x_test, y_test)
print(score*100)
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
y_pred=nb.predict(x_test)
score = nb.score(x_test, y_test)
print(score*100)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=10,random_state=10,max_features=None,min_samples_leaf=15)
tree.fit(x_train, y_train)
y_pred=tree.predict(x_test)
score = tree.score(x_test, y_test)
print(score*100)
```

```python
from sklearn.svm import SVC 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
svm=SVC(kernel="linear",C=0.025)
svm.fit(x_train,y_train)
y_pred=svm.predict(x_test)
score = svm.score(x_test, y_test)
print(score*100) 
```

```python
from sklearn.ensemble import RandomForestClassifier
rfm = RandomForestClassifier(n_estimators=70,oob_score=True,n_jobs=1,random_state=101,max_features=None,min_samples_leaf=30)
rfm.fit(x_train,y_train)
y_pred=rfm.predict(x_test)
score = rfm.score(x_test, y_test)
print(score*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
score = knn.score(x_test, y_test)
print(score*100)
```
