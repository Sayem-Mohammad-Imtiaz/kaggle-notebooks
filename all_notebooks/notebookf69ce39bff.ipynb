```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("../input/historical-us-president-physical-data-more/Historical Presidents Physical Data (More).csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop(['height_in','weight_lb','birth_day','birth_month','birth_year','death_day','death_month','death_year'],axis=1,inplace=True)
```

```python
plt.figure(figsize=(35,12))
sns.barplot(data=data , x= 'name', y = 'death_age')
plt.xticks(rotation=90)
plt.show()
```
