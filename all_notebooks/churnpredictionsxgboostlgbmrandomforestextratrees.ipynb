```python
bank_churners = bank_churners.iloc[:, :-2]
bank_churners.head()
sns.pairplot(bank_churners, hue = 'Attrition_Flag')
```

```python
fig, ax = plt.subplots(figsize = (10, 10))
sns.heatmap(bank_churners.corr(), ax = ax, annot = True)
```

```python
churn_dict = {'Existing Customer':0, 'Attrited Customer':1}
bank_churners['churn'] = bank_churners['Attrition_Flag'].apply(lambda x: churn_dict[str(x)] )
bank_churners.head()
```

```python
y = bank_churners['churn']
bank_churners = bank_churners.drop(['churn','Attrition_Flag'], axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(bank_churners, y,
                                   test_size = 0.33, stratify = y, random_state = 42)
```

```python
le = LabelEncoder()
le.fit(X_train['Gender'])
X_train['Gender'] = le.transform(X_train['Gender'])
X_test['Gender'] = le.transform(X_test['Gender'])
le = LabelEncoder()
le.fit(X_train['Education_Level'])
X_train["Education_Level"] = le.transform(X_train['Education_Level'])
X_test['Education_Level'] = le.transform(X_test['Education_Level'])
le = LabelEncoder()
le.fit(X_train['Marital_Status'])
X_train["Marital_Status"] = le.transform(X_train['Marital_Status'])
X_test['Marital_Status'] = le.transform(X_test['Marital_Status'])
le = LabelEncoder()
le.fit(X_train['Income_Category'])
X_train["Income_Category"] = le.transform(X_train['Income_Category'])
X_test['Income_Category'] = le.transform(X_test['Income_Category'])
le = LabelEncoder()
le.fit(X_train['Card_Category'])
X_train["Card_Category"] = le.transform(X_train['Card_Category'])
X_test['Card_Category'] = le.transform(X_test['Card_Category'])
```

```python
X_train = X_train.drop('CLIENTNUM', axis = 1)
X_test = X_test.drop('CLIENTNUM', axis = 1)
```

```python
xgb_model = XGBClassifier(n_estimators = 266, learning_rate = 0.05, max_depth = 6,
                         subsample = 0.9, colsample_bytree = 0.7)
```

```python
lgbm_model = LGBMClassifier(n_estimators = 315, learning_rate = 0.1, num_leaves = 37,
                           colsample_bytree = 0.8)
```

```python
rf_model = RandomForestClassifier(n_estimators = 201, max_features = 0.6)
```

```python
etc_model = ExtraTreesClassifier(n_estimators = 250, max_features = None)
```

```python
folds = StratifiedKFold(n_splits = 5)
```

```python
score = np.mean(cross_val_score(etc_model, X_train, y_train, scoring = 'f1', cv = folds))
score
```

```python
xgb_model.fit(X_train, y_train)
xgb_importances = xgb_model.feature_importances_
xgb_predictions = xgb_model.predict(X_test)
```

```python
xgb_importances = pd.Series(xgb_importances, index = X_train.columns)
xgb_importances.plot(kind = 'bar')
```

```python
lgbm_model.fit(X_train, y_train)
lgbm_importances = lgbm_model.feature_importances_
lgbm_predictions = lgbm_model.predict(X_test)
```

```python
lgbm_importances = pd.Series(lgbm_importances, index = X_train.columns)
lgbm_importances.plot(kind = 'bar')
```

```python
rf_model.fit(X_train, y_train)
rf_importances = rf_model.feature_importances_
rf_predictions = rf_model.predict(X_test)
```

```python
rf_importances = pd.Series(rf_importances, index = X_train.columns)
rf_importances.plot(kind = 'bar')
```

```python
etc_model.fit(X_train, y_train)
etc_importances = etc_model.feature_importances_
etc_predictions = etc_model.predict(X_test)
```

```python
etc_importances = pd.Series(etc_importances, index = X_train.columns)
etc_importances.plot(kind = 'bar')
```

```python
recall_xgb = recall_score(y_test, xgb_predictions)
recall_xgb
```
