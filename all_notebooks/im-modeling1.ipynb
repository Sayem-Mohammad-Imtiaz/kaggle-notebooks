Функция генерации точек

```python
a1 = 0
a2 = 3
b1 = 0
b2 = 2.5
def gen (n):
    normalized = []
    for i in range(n):
        x = random.uniform(a1, a2)
        y = random.uniform(b1, b2)
        normalized.append([x,y])
    return normalized
```

```python
def check (x , y):
    if y < 0:
        return False
    if x > 3:
        return False
    if y > np.tan(x) and x < math.pi/2 or y < np.tan(x) and x > math.pi/2:
        return False
    if x != 0 and y > 1/x:
        return False
    return True
```

Функция для получения результата

```python
def res (m , p):
    normalized = []
    for i in range(m):
        dots = gen(p)
        col = 0
        for j in range(p):
            if check(dots[j][0], dots[j][1]):
                col = col + 1
        normalized.append(col/p*100)
    return normalized
```

500 точек

```python
result = res (5 , 500)
result
```
