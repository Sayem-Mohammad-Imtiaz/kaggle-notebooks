<h3>1 Introduction</h3>
<p style='text-align: justify;'>The industry of international space exploration has greatly advanced since World War II and the Nuclear Arms Race. Following the Soviet Unionâ€™s launch of Sputnik, the space race ensued, with countries developing satellites, rockets, and other technologies to further explore outer space. With this rapid growth of the new Space Industry comes the question of whether the privatization of outer space is fair and equitable for all nations. 
    
<p style='text-align: justify;'>Prior to 2004 private space travel was considered as illegal. In 2004 Commercial Space Launch Amendments Act was introduce to ripped away the limitations of space travel and allowed private resources to enter into spcae exploration. With the successful takeoff of the SpaceX Falcon 9 rocket on 30 May 2020, marked a new era in human spaceflight and new era in the privatisation of space. </p>

<h4>3.8 Comparision of Succes rate between Private and States</h4>

In this section we compare the success rate between state and private launches

```python
suc_df = df[df['Private or State Run']=='P']

tot = suc_df['Status Mission'].count()
print("Total missions from Private = ", tot)
suc = suc_df[suc_df['Status Mission']=='Success']['Status Mission'].count()
print("Success missions from Private = ", suc)
print("Success rate of Private = ", round(100 * suc/tot,1),'%')
```

```python
suc_df = df[df['Private or State Run']=='S']

tot = suc_df['Status Mission'].count()
print("Total missions from State = ", tot)
suc = suc_df[suc_df['Status Mission']=='Success']['Status Mission'].count()
print("Success missions from State = ", suc)
print("Success rate of State = ", round(100 * suc/tot,1),'%')
```

<h3>4 Hypothesis Testing</h3>

<h4>4.1 Potential Hypotheses </h4>

1) Private sector is more succesfull than state space launch missions

2) USA has more succesful mission percentage than Russia

3) SpaceX has more failures than NASA

<h4>4.2 Significant Test: Hypothesis 1 </h4>

<b>Hypothesis: </b> Private sector succes is less than 90%.

<b>Alternative Hypothesis: </b> Assuming private sector has more succesful percentage (above 90%) compared to state.

We have two outcomes, one with probability (p) and the other with probability (1-p) of succes, so we know the binomial distribution is the right tool for this analysis.

```python
from scipy.stats import binom
prob = 1 - binom.cdf(89, 100, 0.5)

print(str((prob*100,))+"%")
```

The probability of getting 90 or more correct is about ~0%. This is less than 5%, so we can reject the null and conclude that private space launches are more succesfull than state launches.
