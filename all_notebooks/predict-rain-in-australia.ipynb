```python
import numpy as np 
import pandas as pd
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv("../input/weather-dataset-rattle-package/weatherAUS.csv")
df.info()
```

First, do some preprocessing, dropping RISK_MM, this is highly relevant to "RainTomorrow". and location seems not necessary. Coverting date to date format and extract their seasons.

```python
df.drop(["RISK_MM","Location"], axis=1, inplace = True)
df.head()
```

```python
df.Date= pd.to_datetime(df.Date)
df.Date =pd.qcut(df.Date.dt.month,labels=["Summer","Autumn","Winter","Spring"],q =4)
```

Basic feature engineering, standardizing scalar, encoding labels and applying PCA to reduce redundent features.

```python
from sklearn.preprocessing import LabelEncoder,StandardScaler
```

```python
stander = StandardScaler()
for i in df.columns:
    if df[i].dtype.name == "float64":
        a = df[i].values.reshape((-1,1))
        df[i] = stander.fit_transform(a)
df.head()
```

```python
encoder = LabelEncoder()

for i in df.columns:

    if df[i].dtype.name == "object" or df[i].dtype.name == "category":
        df[i].replace(to_replace=np.nan, value ="N/A", inplace=True)
        df[i]=encoder.fit_transform(df[i])
```

```python
df.head()
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
pca = PCA(tol = 0.95)
```

```python
data = df.iloc[:,0:-1]
target = df.iloc[:,-1]
```

```python

for i in data.columns:
    data[i].replace(to_replace= np.nan,value=df[i].mean(),inplace = True)

df.isna().sum()
```

```python
data = pca.fit_transform(data)
x_train, x_test , y_train, y_test = train_test_split(data, target, test_size = 0.25) 
```

Try Logistic Regression first.

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
LR.fit(x_train,y_train)
```

```python
LR.predict(x_test)
```

Score for test sets

```python
LR.score(x_test,y_test)
```
