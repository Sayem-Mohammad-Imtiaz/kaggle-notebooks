# Importing dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Loading data

```python
data = pd.read_csv('../input/house-prices-advanced-regression-techniques/train.csv')
test_data = pd.read_csv('../input/house-prices-advanced-regression-techniques/test.csv')
print(data.head(4))
print(test_data.head(4))
```

# Data Cleaning

```python

arr=[]
for i in data.columns:
    if data[i].isnull().sum() != 0 :
        arr.append(i)
print(arr)
data.drop(arr,inplace=True,axis=1)
data.columns
```

# Creating Models

```python
ar=[]
for i in data.columns :
    if data[i].dtypes == 'object' :
        ar.append(i)
ar
data.drop(ar,inplace=True,axis=1)

arr = ['BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF','BsmtFullBath',
       'BsmtHalfBath','GarageCars', 'GarageArea',]
data.drop(arr,inplace=True,axis=1)
data.dtypes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn import metrics
X = data.drop(['SalePrice','Id'],axis=1)
y = data.SalePrice
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=4)
```

```python
from sklearn.linear_model import LinearRegression
model_lr = LinearRegression()
model_lr.fit(X,y)
model_lr.score(X,y)
```

```python
y_pred = model_lr.predict(X_test)
metrics.r2_score(y_pred,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model_rf = RandomForestRegressor(n_estimators =100 ,random_state=4)
model_rf.fit(X,y)
y_pred_rf = model_rf.predict(X_test)
model_rf.score(X,y)
```

```python
metrics.r2_score(y_pred_rf,y_test)
```
