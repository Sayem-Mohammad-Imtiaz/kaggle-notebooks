```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly
```

# Create a COVID-19 world vaccination map

```python
df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv', header=0)
print(df.columns)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
fig = px.choropleth(df, locations='country', locationmode='country names', color='people_fully_vaccinated_per_hundred', hover_name='country', animation_frame='date', color_continuous_scale='rainbow', range_color=(0.0, 40.0))
fig.update_layout(title_text='Proportion people fully vaccinated (%)', title_x=0.5)
fig.show()

plotly.offline.plot(fig, filename='vaccination.html')
```

# Create a world map of earthquakes

```python
df = pd.read_csv('../input/earthquakes-20181127/earthquakes.csv', header=0)
print(df.columns)
#print(df['time'])
df.index = pd.to_datetime(df['time'])
#print(df.index)
df['time'] = df.index.strftime('%Y-%m-%d %H:00:00')
#print(df['time'])
df = df.sort_index()
df['mag_exp'] = df['mag']**3  # make the larger magnitude earthquakes visible with larger dots
fig = px.scatter_geo(df, lat='latitude', lon='longitude', color='mag', animation_frame='time', color_continuous_scale='rainbow', range_color=(0.0, 8.0), size='mag_exp')
fig.update_layout(title_text='Earthquakes 2018-11-27 \nAnchorage 7.1', title_x=0.5)
fig.show()

#plotly.offline.plot(fig, filename='vaccination.html')
```
