```python
# import useful libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset

data= pd.read_csv("../input/breast-cancer-wisconsin-data/data.csv")
data
```

```python
data.info()
```

```python
# lets drop the two columns that doesnt contribute to the data
column = ['id', 'Unnamed: 32']
new_data = data.drop(column, axis = 1)
```

```python
new_data.isnull().sum().sum()
```

```python
new_data.describe()
```

```python
# visualizing the diagnosis column to be sure it isnt imbanced
sns.countplot(new_data['diagnosis'])
```

```python
# visulaize the correlation between the dataset
plt.figure(figsize= (10,10))
sns.heatmap(new_data.corr())
```

```python
# split into target and features

X = new_data.drop('diagnosis', axis = 1)
y = new_data['diagnosis']
```

```python
# splitting the data set into train and test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size= 0.3, random_state = 42)
```

```python
X_train.shape
```

```python
X_test.shape
```

### Build the classification model

```python
# logistic regression model
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.svm import LinearSVC

# using a pipeline to make the code neater, with a feature selection and classigication model
clf = Pipeline([
  ('feature_selection', SelectFromModel(LinearSVC(C=0.01, penalty="l1", dual=False))),
  ('classification', LogisticRegression())
])
clf.fit(X, y)
```

```python
pred= clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print(classification_report(y_test, pred))
```

```python
# view the classification result with confusion matrix
cm = confusion_matrix(y_test, pred)
cm
```

```python
# visualize the confusion matrix with a heatmap

sns.heatmap(cm, annot= True, fmt= 'g')
```

Based on the model, the accuracy is about 97%. There is a false positive of about 3 and False negatives of 2.
Generally, the model performed well.
