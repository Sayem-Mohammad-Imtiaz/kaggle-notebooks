```python
import pandas
AZhousing_file_data='/kaggle/input/arizona-houses-2021/AZhousingData.csv' #文件路径
AZhousing_data=pandas.read_csv(AZhousing_file_data)  #读取文件 
AZhousing_data.columns  #查看表头
AZhousing_data=AZhousing_data.dropna(axis=0) #删除缺失值
y=AZhousing_data.Price  #赋值价格  
AZhousing_features=['beds','baths','sqft']  #特征
X=AZhousing_data[AZhousing_features]  #重新给x赋值  训练数据
X.describe()  #数据摘要

from sklearn.tree import DecisionTreeRegressor  #导入模块的方法
AZhousing_model=DecisionTreeRegressor()  #定义模型
AZhousing_model.fit(X,y)  #拟合

from sklearn.metrics import mean_absolute_error  #导入模块的方法
predicted_home_price=AZhousing_model.predict(X)  #变量承载数据集中通过x特征得到的房屋预测值 
mean_absolute_error(y,predicted_home_price)#  将目标值y（原始房价值）和刚才得到的预测房价值交给函数对比 会得到平均误差值

from sklearn.model_selection import train_test_split  #导入分割模块
train_X,val_X,train_y,val_y=train_test_split(X,y,random_state=20)  #随机分割 
AZhousing_model=DecisionTreeRegressor()#定义模型
AZhousing_model.fit(train_X,train_y)#拟合数据
verification=AZhousing_model.predict(val_X)#预测价格结果
print(mean_absolute_error(val_y,verification)) #输出绝对误差值 对照参数是目标值val——y
```
