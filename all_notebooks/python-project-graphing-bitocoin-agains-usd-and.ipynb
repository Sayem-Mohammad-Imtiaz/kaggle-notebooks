## Loading the CSV data


```python
import datetime
def dateparse (time_in_secs):    
    return datetime.datetime.fromtimestamp(float(time_in_secs))

btc = pd.read_csv('../input/historical-cryptocurrencies-price-krakencom/XBTUSD_60.csv', parse_dates=['Timestamp'], date_parser=dateparse,  names=['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume', 'dunno']) 
eth = pd.read_csv('../input/historical-cryptocurrencies-price-krakencom/ETHUSD_60.csv', parse_dates=['Timestamp'], date_parser=dateparse,  names=['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume', 'dunno']) 
storj = pd.read_csv('../input/historical-cryptocurrencies-price-krakencom/STORJUSD_60.csv', parse_dates=['Timestamp'], date_parser=dateparse,  names=['Timestamp', 'Open', 'High', 'Low', 'Close', 'Volume', 'dunno']) 
```

# Plot of Cryptocurrencies prices in USD on a single graph

```python
plt.figure(figsize=(30,10))
plt.plot(btc['Timestamp'], btc['Close'], label="BTC in USD")
plt.plot(eth['Timestamp'], eth['Close'], label="ETH in USD")
plt.plot(storj['Timestamp'], storj['Close'], label="STORJ in USD")
plt.legend()
plt.show()
```

## Calculating percent change and log of returns

```python
btc['pct_change'] = btc['Open'].pct_change()
btc['log_return'] = np.log1p(btc['Open'].pct_change())

eth['pct_change'] = eth['Open'].pct_change()
eth['log_return'] = np.log1p(eth['Open'].pct_change())

storj['pct_change'] =storj['Open'].pct_change()
storj['log_return'] = np.log1p(storj['Open'].pct_change())
```

## Plotting log of returns on subplots

```python
fig, axis = plt.subplots(3, figsize=(50,15), sharex=True)
fig.suptitle('Log returns comparison of selected Crypto')
axis[0].plot(btc['Timestamp'], btc['log_return'], label="log return of BTC", )
axis[0].legend()

axis[1].plot(eth['Timestamp'], eth['log_return'], label="log return of ETH")
axis[1].legend()

axis[2].plot(storj['Timestamp'], storj['log_return'], label="log return of STORJ")
axis[2].legend()


```

```python
fig, axis = plt.subplots(3, figsize=(50,15), sharex=True)
fig.suptitle('Log returns comparison of selected Crypto')
axis[0].plot(btc['Timestamp'], btc['pct_change'], label="log return of BTC", )
axis[0].legend()

axis[1].plot(eth['Timestamp'], eth['pct_change'], label="log return of ETH")
axis[1].legend()

axis[2].plot(storj['Timestamp'], storj['pct_change'], label="log return of STORJ")
axis[2].legend()


```

This means that there are no consistent patterns in the data that allow you to model stock prices over time near-perfectly.
