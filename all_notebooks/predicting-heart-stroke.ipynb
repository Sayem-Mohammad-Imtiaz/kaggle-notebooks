## Visualizing the DATA

```python
data = df_nona
```

```python
data.columns
```

```python
sns.countplot(x="gender", data = data)
```

```python
data[data['gender']=="Other"]
```

```python
features = data.columns

def plot_countplot(features):
    for i in features:
        plt.figure(figsize=(15,5))
        sns.countplot(x=i, data=data)
        plt.show()
```

```python
features = features.drop("age")
```

```python
plot_countplot(features)
```

- Male count is between 10,000 and 12,500 and females are 17,500 and others are 6
- less than 5000 or nearly 2500 people have hyper tension in total of 29,000 people
- and one-third of 5000 people are having heart disease
- nealry 7500 people are not married out of 29000 people
- less than 1000 people have never worked (including children)
- Around 50% of people have never smoked
- very uneven data can be seen as stroke patient are very less compared to people who doesnt have stroke
