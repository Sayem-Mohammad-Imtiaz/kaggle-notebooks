```python
#import libraries

#Basic libraries
import pandas as pd
import numpy as np

#Libraries for visualization
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#Load data


gold_rates_1979_2021 = pd.read_csv('../input/monthly-gold-prices/1979-2021.csv')
gold_rates_1990_2021 = pd.read_csv('../input/monthly-gold-prices/1990-2021.csv')
```

## EDA of monthly prices from 1979 to 2021

```python
gold_rates_1979_2021.head()
```

```python
#Let's check info about data set
gold_rates_1979_2021.info()
```

```python
#Describe data set
gold_rates_1979_2021.describe()
```

### Missing values

```python
#Check missing values
gold_rates_1979_2021.isnull().any()
```

```python
#Get info about missing values
missing_months = gold_rates_1979_2021.isnull().sum()

missing_percent =  gold_rates_1979_2021.isna().sum() / len(gold_rates_1979_2021) *100
missing_percent = missing_percent.round(2)

missing_years = gold_rates_1979_2021.isna().sum() / 12
missing_years = missing_years.round(2)

#create a table 
table = pd.concat([missing_percent,missing_months,missing_years], axis = 1)
table.columns = ['missing_percent', 'missing_months', 'missing_years']
table = table.drop(table[(table.missing_percent == 0) & (table.missing_months == 0) & (table.missing_years == 0)].index)

table
```

### Data visualization

```python
#Add new columns (Year and Month)
gold_rates_1979_2021['Year'] =gold_rates_1979_2021['Date'].apply(lambda x: x.split("-")[2])
gold_rates_1979_2021['Month'] = gold_rates_1979_2021['Date'].apply(lambda x: x.split("-")[1])
```

```python
plt.figure(figsize = (20, 10))
sns.lineplot(x='Year', y='United States(USD)',  hue='Month', style='Month',
             data=gold_rates_1979_2021)
plt.xticks(rotation = 90)
plt.ylabel('Price(USD)')
plt.title('Gold prices between 1979 and 2021 - United States(USD)')
plt.show()
```

```python
plt.figure(figsize = (20, 10))
sns.lineplot(x='Year', y='Europe(EUR)',  hue='Month', style='Month',
             data=gold_rates_1979_2021)
plt.xticks(rotation = 90)
plt.ylabel('Price(EUR)')
plt.title('Gold prices between 1979 and 2021 - Europe(EUR)')
plt.show()
```
