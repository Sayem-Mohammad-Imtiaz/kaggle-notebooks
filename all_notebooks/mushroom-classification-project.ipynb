<p> <font face="tahoma" size="5" color="Brown">Machine Learning Models </p>

<p> <font face="tahoma" size="4" color="Brown">Split the data into train dataset and test data set </p>

```python
le = LabelEncoder()

df_encoded = mushroom.apply(le.fit_transform)

y = df_encoded.iloc[:, :1].values
X = df_encoded.iloc[:, 1:].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

accryArray=[]
```

<p> <font face="tahoma" size="4" color="Brown">K-Nearest Neighbors Method </p>

```python
for n in range(2,20):
    classifier = KNeighborsClassifier(n_neighbors=n)
    classifier.fit(X_train, y_train.ravel())
    y_pred = classifier.predict(X_test)
    predict = metrics.accuracy_score(y_test, y_pred)
    #print("For n=",n," Accuracy:",predict)
    accryArray.append(predict)

plt.scatter([i for i in range(2,20)],accryArray)
plt.plot([i for i in range(2,20)],accryArray)
plt.show()
```
