```python
import json
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

df = pd.read_csv('../input/instagram-fake-and-real-accounts-dataset/final-v1.csv')

corrMatrix = df.corr()

## show correlation matrix
sn.heatmap(corrMatrix, annot=True)
F = plt.gcf()
Size = F.get_size_inches()
F.set_size_inches(Size[0]*2, Size[1]*2, forward=True) # Set forward to True to resize window along with plot in figure
plt.show()
```

Parameter 'is_fake' has important correlation with has_external_url. Also is_bussiness_account and username that contains number should be noticed.
