```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline
Calls = pd.read_csv('/kaggle/input/911-calls/911.csv')
```

```python
Calls['twp'].value_counts().head(5)
```

```python
Calls['title'].nunique()
```

```python
Calls['Reason'] = Calls['title'].apply(lambda title: title.split(':')[0])
```

```python
Calls['Reason'].value_counts().head(3)
```

```python
sns.countplot(Calls['Reason'],data = Calls,palette ='viridis')
```

```python
type(Calls['timeStamp'].iloc[0])
```

```python
Calls['timeStamp'] = pd.to_datetime(Calls['timeStamp'])
```

```python
Calls['Hour'] = Calls['timeStamp'].apply(lambda time: time.hour)
Calls['Month'] = Calls['timeStamp'].apply(lambda time: time.month)
Calls['Day of Week'] = Calls['timeStamp'].apply(lambda time: time.dayofweek)
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
Calls['Day of Week'] = Calls['Day of Week'].map(dmap)
```

```python
sns.countplot(x='Day of Week',data=Calls,hue='Reason',palette='viridis')

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.5)
```

```python
sns.countplot(x='Month',data=Calls,hue='Reason',palette='rocket')

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.5)
```

```python
byMonth = Calls.groupby('Month').count()
byMonth.head()
```

```python
byMonth['e'].plot()
```

```python
sns.lmplot(x='Month',y='e',data=byMonth.reset_index())
plt.ylabel('Calls per month')
```

```python
Calls['Date'] = Calls['timeStamp'].apply(lambda t : t.date())
```

```python
Calls.groupby('Date').count()['twp'].plot(figsize = (15,3))
plt.ylabel('Calls per month')
```

```python
Calls[Calls['Reason']=='Traffic'].groupby('Date').count()['twp'].plot(figsize = (15,3))
plt.ylabel('Calls per month')
plt.title('Traffic')
plt.tight_layout()
```

```python
Calls[Calls['Reason']=='EMS'].groupby('Date').count()['twp'].plot(figsize = (15,3))
plt.ylabel('Calls per month')
plt.title('EMS')
plt.tight_layout()
```

```python
Calls[Calls['Reason']=='Fire'].groupby('Date').count()['twp'].plot(figsize = (15,3))
plt.ylabel('Calls per month')
plt.title('Fire')
plt.tight_layout()
```

```python
dayHour = Calls.groupby(by=['Day of Week','Hour']).count()['Reason'].unstack()
dayHour.head(7)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayHour,cmap='magma')
```

```python
plt.figure(figsize=(12,6))
sns.clustermap(dayHour,cmap='viridis', figsize = (12,5))
```

```python
dayMonth = Calls.groupby(by=['Day of Week','Month']).count()['Reason'].unstack()
dayMonth.head(7)
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(dayMonth,cmap='magma')
```

```python
sns.clustermap(dayMonth,cmap='viridis', figsize = (12,5))
```
