# Imports

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Opening CSV
df = pd.read_csv('../input/brasilian-houses-to-rent/houses_to_rent_v2.csv')
df.head() # First 5 lines
```

```python
#nÂº (lines, columns)
df.shape
```

```python
# General information
df.info()
```

```python
# Cheking correlation between features
fig, ax = plt.subplots(figsize=(10,8))
sns.heatmap(df.corr(), vmin=-1, vmax=1, cmap=sns.diverging_palette(20, 220, as_cmap=True), annot=True)
```

  # Which features chosse?
  The correlation grafic shows high correlation between total (R$) vs hoa
  > The following features were selected: rooms, parking space and hoa

```python
# Features

previsores = df.iloc[:,[3,4,8]].values

# Target
classes = df.iloc[:,12].values

```

```python
# Separating (x,y) train and (x,y) test
x_train, x_test, y_train, y_test = train_test_split(previsores, classes, test_size = 0.30,random_state = 0)
```

# Creating model
Here is the choice of linear regression to try to predict rantal prices houses

```python
# calling the linear regression
modelo = LinearRegression()

#training de model
modelo.fit(x_train,y_train)
```

```python
#
modelo.predict
```

```python
# point where the axis x crosses the axis y
modelo.intercept_
```

```python
# Linear coef of axis x1, x2 and x3
modelo.coef_
```

```python
# f(x1,x2,x3) = coefx1 +coefx2 + coefx3 + b
modelo.predict(x_test)
```

# Validation

```python
# prev is the predicted target
prev = modelo.predict(x_test)
```

```python
# calculating the mean absolute error
mean  = mean_absolute_error(prev, y_test)
mean
```

```python
df['total (R$)'].mean() # total mean
```

```python
prev.mean() # mean of predicted data
```

Here you can see that the average **df['total (R$)']** is close to the **prev** average
