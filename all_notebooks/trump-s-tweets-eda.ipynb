```python
#import libraries
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#create dataframe
exprezdf = pd.read_csv("../input/all-trumps-twitter-insults-20152021/trump_insult_tweets_2014_to_2021.csv")
exprezdf.head(10)
```

```python
#Check data shape
exprezdf.shape
```

```python
#Check null values
exprezdf.isnull().sum()
```

```python
#Drop null values
exprezdf.dropna(inplace=True)
exprezdf.shape
```

```python
#Let's look at insults
exprezdf.insult.unique().shape
```

```python
insultcounts = exprezdf.target.value_counts()
insultcounts
```

```python
#Pull insults and count of each into separate dataframe
insults = exprezdf.groupby(['insult'])['Unnamed: 0'].count().reset_index(name='count').sort_values(by='count',ascending=False)
insults.head()
```

```python
#Pull out top insults from insult dataframe
top_insults = insults.groupby('insult').aggregate(sum).sort_values('count', ascending=False).index
number=15
fav_insults = insults.loc[insults.insult.isin(top_insults[:number])]
fav_insults.head(15)
```

```python
# make barplot and sort bars in descending order
plt.figure(figsize=(10,10))
sns.barplot(x='count', 
            y="insult", 
            data=fav_insults, 
            order=fav_insults.sort_values('count',ascending = False).insult)
# set labels
plt.xlabel("count", size=15)
plt.ylabel("insult", size=15)
plt.title("Trump's Favorite Insults", size=18)
plt.tight_layout()
plt.show()
```
