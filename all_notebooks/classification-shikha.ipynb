```python
model = Sequential([
    Dense(units=8, input_shape=(10,), activation='relu'),
    Dense(units=8, activation='relu'),
    Dense(units=8, activation='relu'),
    Dense(units=2, activation='sigmoid' )
])
```

```python
model.compile(
optimizer=Adam(learning_rate=0.0001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
model.fit(
x=X_train,
y=y_train,
    batch_size=10,
    epochs=30,
    shuffle=True,
    verbose=2
)
```
