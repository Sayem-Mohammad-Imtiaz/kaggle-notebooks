```python
df = pd.read_csv("/kaggle/input/ipl-20082020-batsman-vs-bowler/Batsmen vs bowler IPL 2008 to 2020.csv")
df
```

```python
df.info()
```

```python
# Remove whitespace from "bowler" columns
df.columns = ['batsman', 'bowler', 'runs_scored', 'balls_faced']
df.columns
```

```python
# Calculate the strike rate
df["strike_rate"] = round(df["runs_scored"] / df["balls_faced"] * 100, 2)
df[:10]
```

```python
# Let's check out almost everyone's favourate bastmen
abd = df["batsman"].str.contains("de Villiers")
df[abd]
```

```python
# Let's see the ten bowlers AB de Villiers scored the most runs from
df[abd].sort_values("runs_scored", ascending = False)[:10]
```

```python
# Lets also check out AB's strike rates. But lets make the criteria : minimum 30 runs scored
min_runs_scored = df["runs_scored"] > 29
df[abd & min_runs_scored].sort_values("strike_rate", ascending = False)[:10]
```

```python
df.info()
```

```python
# Create new column
df["batsman_vs_bowler"] = df["batsman"] + " vs " + df["bowler"]
df
```

```python
df = df[["batsman_vs_bowler", "runs_scored", "balls_faced", "strike_rate"]]
df[:10]
```

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.style
```

```python
# Top 10 most runs against bowler
top_ten = df.sort_values("runs_scored",ascending = False)[:10]
top_ten
```

```python
matplotlib.style.use("bmh")
ax = top_ten.plot(figsize = (14,7), kind = "scatter", x = "runs_scored", y = "strike_rate")

for i, txt in enumerate(top_ten["batsman_vs_bowler"]):
    ax.annotate(txt,(top_ten.runs_scored.iat[i],top_ten.strike_rate.iat[i]))

plt.title("IPL Batsman vs bowler : Most runs with strike rate")
plt.show()
```
