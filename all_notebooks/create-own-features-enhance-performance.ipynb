# Generating new features

we can generate new features by taking product of two or more features, (with these following procedure):-

1. The feature should have high varience, which can make high differnce in prediction. #(eg. like in [pH vs quantity] graph pH have low varience, if removing pH features it doesn't effect accuracy.

2. If two or more features are making nearly equal effect on quality of wine. #(eg. as you can see the graph of (alcohol vs quality) and (catric acid vs quality), incrementing in the value of both leads the quality of wine between 6-8, the multiplication of both can give good new features with the probablity of both feature, which help in prediction more accurately.)  

```python
length = len(x[:,0])
x_1 = (x[:,5])*x[:,7] # Multiplying free sulfer dioxide and density
x_1 = np.reshape(x_1,(length,1)) # reshaping (length,1)
x_2 = (x[:,7])*x[:,3] # multiplying residual sugar and density
x_2 = np.reshape(x_2,(length,1)) 
x_3 = x[:,2]*x[:,10] # alcohol and cetric acid
x_3 = np.reshape(x_3,(length,1))

x = np.hstack((x,x_3,x_1,x_2)) # concatinating all 3 new features 
print(x.shape)
```

```python
#convert result into (0,1) binary form
y = y>=6.5
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)

tune = {'kernel':['rbf'],'gamma':['scale'],'C':[1,10,100,1000]}

clf = GridSearchCV(svm.SVC(),tune)
clf.fit(x_train,y_train)

print('accuracy score = ',clf.score(x_test,y_test))
print(clf.best_params_)
```

```python
print(classification_report(y_test, clf.predict(x_test)))
```

```python
print(confusion_matrix(y_test, clf.predict(x_test)))
```

```python

random_f=RandomForestClassifier(n_estimators=230)
random_f.fit(x_train,y_train)

print(random_f.score(x_test,y_test))
```
