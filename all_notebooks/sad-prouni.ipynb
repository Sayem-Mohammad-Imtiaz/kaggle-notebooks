```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import matplotlib.pyplot as plt

import os
print(os.listdir("../input/pro-uni20052009"))
```

```python
# LENDO O ARQUIVO DE ENTRADA E GERANDO UM DATAFRAME DO PANDAS
df = pd.read_csv("../input/pro-uni20052009/PROUNI.csv", delimiter=';')
df.head(5)
```

Começando nossa análise visualizando a quantidade dos três diferentes tipos de bolsas ao longo dos anos.

```python
# SELECIONANDO AS TRÊS COLUNAS DE INTERESSE
bolsa_regiao_ano = df[["ID","REGIAO_BENEFICIARIO_BOLSA","ANO_CONCESSAO_BOLSA"]]

# AGRUPANDO POR REGIÃO E ANO, E AGREGANDO EM UMA NOVA COLUNA UMA CONTAGEM DE BOLSAS POR REGIÃO E ANO
bolsa_regiao_ano = bolsa_regiao_ano.groupby(["REGIAO_BENEFICIARIO_BOLSA","ANO_CONCESSAO_BOLSA"]).agg({'ID':['count']}).reset_index()

# RENOMEANDO AS COLUNAS
bolsa_regiao_ano.columns = ['Regiao','Ano','Total de Bolsas']

# TRANSFORMANDO EM UMA NOVA TABELA EM QUE AS NOVAS COLUNAS SÃO ['Ano','BOLSA COMPLEMENTAR 25%','BOLSA INTEGRAL','BOLSA PARCIAL 50%']
bolsa_regiao_ano = bolsa_regiao_ano.pivot_table(index='Ano', columns='Regiao', values='Total de Bolsas').fillna(0).reset_index()

# REMOVENDO A COLUNA "Em Branco"
bolsa_regiao_ano.pop("Em Branco")

# PLOTANDO O GRÁFICO DE QUANTIDADE DE BOLSAS AO LONGO DOS ANOS
bolsa_regiao_ano.plot.line(x='Ano')
# bolsa_regiao.head(50)
```

```python
# SELECIONANDO AS DUAS COLUNAS DE INTERESSE
tipo_bolsa_ano = df[["TIPO_BOLSA","ANO_CONCESSAO_BOLSA"]]

# AGRUPANDO POR TIPO DE BOLSAS E ANO, E AGREGANDO EM UMA NOVA COLUNA UMA CONTAGEM DE BOLSAS POR ANO E TIPO
tipo_bolsa_ano = tipo_bolsa_ano.groupby(["TIPO_BOLSA","ANO_CONCESSAO_BOLSA"]).agg({'TIPO_BOLSA':['count']}).reset_index()

# RENOMEANDO AS COLUNAS
tipo_bolsa_ano.columns = ['Tipo de Bolsa','Ano','Total de Bolsas']

# TRANSFORMANDO EM UMA NOVA TABELA EM QUE AS NOVAS COLUNAS SÃO ['Ano','BOLSA COMPLEMENTAR 25%','BOLSA INTEGRAL','BOLSA PARCIAL 50%']
tipo_bolsa_ano = tipo_bolsa_ano.pivot_table(index='Ano', columns='Tipo de Bolsa', values='Total de Bolsas').fillna(0).reset_index()

# REORDENANDO AS COLUNAS PARA MELHOR APRESENTAÇÃO
tipo_bolsa_ano = tipo_bolsa_ano.reindex(columns= ['Ano','BOLSA INTEGRAL','BOLSA PARCIAL 50%','BOLSA COMPLEMENTAR 25%'])

# PLOTANDO O GRÁFICO DE QUANTIDADE DE BOLSAS POR ANO
tipo_bolsa_ano.plot.line(x='Ano')
```

```python
# SELECIONANDO AS DUAS COLUNAS DE INTERESSE
bolsa_regiao = df[["ID","REGIAO_BENEFICIARIO_BOLSA"]]

# AGRUPANDO POR TIPO DE BOLSAS E ANO, E AGREGANDO EM UMA NOVA COLUNA UMA CONTAGEM DE BOLSAS POR ANO
bolsa_regiao = bolsa_regiao.groupby(["REGIAO_BENEFICIARIO_BOLSA"]).agg({'ID':['count']}).reset_index()

# RENOMEANDO AS COLUNAS
bolsa_regiao.columns = ['Regiao','Porcentagem de Bolsas']
bolsa_regiao = bolsa_regiao.set_index('Regiao')

# PLOTANDO O GRÁFICO DE PORCENTAGEM DE BOLSAS POR REGIÃO
bolsa_regiao.plot.pie(y='Porcentagem de Bolsas', legend = False, radius = 1.2, autopct='%1.1f%%', pctdistance=0.75, labeldistance=1, title = "Porcentagem de Bolsas por região", ylabel='')
# bolsa_regiao.head(50)
```

```python
# SELECIONANDO AS DUAS COLUNAS DE INTERESSE
bolsa_faixa_etaria = df[["ID","FAIXA_ETARIA_BENEFICIARIO"]]

# AGRUPANDO POR TIPO DE BOLSAS E ANO, E AGREGANDO EM UMA NOVA COLUNA UMA CONTAGEM DE BOLSAS POR ANO
bolsa_faixa_etaria = bolsa_faixa_etaria.groupby(["FAIXA_ETARIA_BENEFICIARIO"]).agg({'ID':['count']}).reset_index()

# RENOMEANDO AS COLUNAS E DEFININDO UM INDEX MELHOR
bolsa_faixa_etaria.columns = ['Faixa Etaria','Total de Bolsas']
bolsa_faixa_etaria = bolsa_faixa_etaria.set_index('Faixa Etaria')

# SELECIONANDO AS MELHORES FAIXAS ETARIAS PARA MELHORAR A VISUALIZAÇÃO NO GRÁFICO
bolsa_faixa_etaria = bolsa_faixa_etaria[1:13].sort_index(ascending=False)

# PLOTANDO O GRÁFICO DE QUANTIDADE DE BOLSAS POR ANO
bolsa_faixa_etaria.plot.barh(title = "Qualtidade de bolsas por faixa etaria", grid = True)
# bolsa_faixa_etaria.head(50)
```

```python
def MaioresCursosPorRegiao(df,regiao):
    # SELECIONANDO AS DUAS COLUNAS DE INTERESSE
    cursos_regiao = df[["ID","NOME_CURSO_BOLSA", "ANO_CONCESSAO_BOLSA"]]

    # SELECIONANDO AS BOLSAS SOMENTE DE UMA REGIÃO
    cursos_regiao = cursos_regiao[df["REGIAO_BENEFICIARIO_BOLSA"]==regiao]

    # SELECIONANDO AS BOLSAS A PARTIR DE 2015
    cursos_regiao = cursos_regiao[cursos_regiao["ANO_CONCESSAO_BOLSA"] >= 2015]

    # AGRUPANDO PELO CURSO, E AGREGANDO EM UMA NOVA COLUNA UMA CONTAGEM DE BOLSAS
    cursos_regiao = cursos_regiao.groupby(["NOME_CURSO_BOLSA"]).agg({'ID':['count']}).reset_index()

    # RENOMEANDO AS COLUNAS
    cursos_regiao.columns = ['Curso','Quantidade de Bolsas']

    # ORDENANDO OS CURSOS PELA MAIOR QUANTIDADE DE BOLSAS
    cursos_regiao = cursos_regiao.sort_values('Quantidade de Bolsas', ascending=False).reset_index(drop=True)

    # DEFININDO O CURSO COMO INDEX
    cursos_regiao = cursos_regiao.set_index('Curso')

    # SELECIONANDO  OS DEZ MAIORES CURSOS
    cursos_regiao = cursos_regiao[:10]
    
    return cursos_regiao
```

```python
REGIAO = "CENTRO-OESTE"
maiores_curso_regiao = MaioresCursosPorRegiao(df,REGIAO)
maiores_curso_regiao.plot.barh(title="Os dez cursos com mais bolsas no " +  REGIAO.capitalize() + " nos últimos cinco anos", xlabel='')
```
