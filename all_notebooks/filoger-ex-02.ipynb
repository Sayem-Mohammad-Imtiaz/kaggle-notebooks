# Q4: missing value

```python
missing_value = df.isnull().values.any()
print('Missing value in the dataset = {}'.format(missing_value))
if missing_value:
    print("There is '{}' missing data in dataset".format((df.isnull().sum()).sum()))
    print("The '{}' column have maximum missing data".format(df.isnull().sum().idxmax()))
else:
    print('There is no missing data in the dataset')
```

# Q5: Deleting the "Company" Columns

```python
df = df.drop('company',axis=1)
```

```python
df.info()
```

# Q6: Declare 5 Country with most passengers

```python
df['country'].value_counts()[:5]
```

# Q7: Most payment

```python
max_adr = df['adr'].idxmax()
name = df.iloc[max_adr]['name']
payment = df.iloc[max_adr]['adr']
print("Most payment is by '{}' and the value is = {}".format(name,payment))
```

# Q8: Average of all ADR

```python
Average_of_ADR = np.round(df['adr'].mean(), 2)
print("Average of Adrs: {}".format(Average_of_ADR))
```

# Q9: Avarage of nights staying

```python
df['stay'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
Average_stay = np.round(df['stay'].mean(),2)
print("Average of nights staying: {}".format(Average_stay))
```

# Q12: People have most children

```python
df["Children"] = df["children"] + df["babies"]
tempt = df[["name","Children"]].sort_values(["Children"],ascending=False)
tempt["name"][0:5]
```

# Q13: 3 most telephone code that reserved the hotel

```python
df['code'] = df['phone-number'].str[0:3]
temp = df.groupby('code').count()['name']
most_phone_reserved = temp.sort_values(ascending=False)[0:3]
print("Most telephone number reserved the hotel are:\n{}".format(most_phone_reserved))
```
