```python
cat = []
num = []
for col in df.columns:
    if df[col].dtype=='O':
        cat.append(col)
    else:
        num.append(col)  
        
        
num 
```

```python
plt.style.use('dark_background')
for col in df[num].drop(['Total cases'],axis=1):
    plt.figure(figsize=(8,5))
    plt.plot(df[col].value_counts(),color='Red')
    plt.xlabel(col)
    plt.ylabel('Total recovered')
    plt.tight_layout()
    plt.show()
```

```python
trace1 = go.Box(
    y=df["Total cases"],
    name = 'Total cases',
    marker = dict(color = 'rgb(0,145,119)')
)

trace2 = go.Box(
    y=df["Total recovered"],
    name = 'Total recovered',
    marker = dict(color = 'rgb(5, 79, 174)')
)

data = [trace1, trace2]
layout = dict(autosize=False, width=700,height=500, title='Africa Covid19', paper_bgcolor='rgb(243, 243, 243)', 
              plot_bgcolor='rgb(243, 243, 243)', margin=dict(l=40,r=30,b=80,t=100,))

fig = dict(data=data, layout=layout)
iplot(fig)
```
