<h3>Now let's see which factor remain constant during the whole timeperiod</h3>

```python
colms=pd.DataFrame()
colms['colm']=[]
colms['count']=[]
```

```python
a=pd.DataFrame(d2015.columns)
b=pd.DataFrame(d2016.columns)
c=pd.DataFrame(d2017.columns)
d=pd.DataFrame(d2018.columns)
e=pd.DataFrame(d2019.columns)
f=pd.DataFrame(d2020.columns)
```

```python
colms=pd.concat([a,b,c,d,e,f,colms])
colms['colm']=colms[0]
colms.drop([0,'count'],axis=1,inplace=True)
```

```python
colms.head()
```

```python
colms['colm'].value_counts()[0:10]
```

```python
plt.figure(figsize=(6,10))
sns.barplot(x=colms['colm'].value_counts(),y=colms['colm'].value_counts().index, palette="RdBu_r")
```
