```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../input/covid19-latest-data-kerala/covid_data_kerala.csv')
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.isnull().any()
```

```python
df['Confirmed']=df['Confirmed'].fillna(df['Confirmed'].median())
```

```python
df.isnull().sum()
```

```python
df['Recovered']=df['Recovered'].fillna(df['Recovered'].median())
```

```python
df['Deceased']=df['Deceased'].fillna(df['Deceased'].median())
```

```python
df.isnull().sum()
```

```python
sns.pairplot(df)
```

```python
x= df['Date']
y = df['Confirmed']
plt.plot(x,y)
plt.xticks(rotation=15)
plt.show()
```

```python
x = df['Date']
y = df['Recovered']
plt.plot(x,y)
plt.show()
```
