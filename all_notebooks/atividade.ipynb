```python
data = pd.read_csv('../input/german-house-prices/germany_housing_data_14.07.2020.csv')
data = data.drop('Unnamed: 0', axis=1)
data = data.drop_duplicates()
data = data.dropna()
data
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
import matplotlib.pyplot as plt
```

```python
a = ['Price',
 'Living_space',
 'Lot',
 'Usable_area',
 'Rooms',
 'Bedrooms',
 'Bathrooms',
 'Floors',
 'Year_built',
 'Year_renovated',
 'Energy_consumption',
 'Garages']
```

```python
data = data.drop(index = 1499)
```

```python
f = plt.figure(figsize=(19, 15))
plt.matshow(data.corr(), fignum=f.number)
plt.xticks(range(len(a)), a, fontsize=14, rotation=45)
plt.yticks(range(len(a)), a, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
```

```python
X = data.drop('Price', axis=1)
X = data[a]
y = data.Price
print(len(X), len(y))
```

```python
# Split dataset into training set and test set
#https://stackabuse.com/random-forest-algorithm-with-python-and-scikit-learn/
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=200, random_state=0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
result = pd.DataFrame({'Pred': y_pred, 'Teste':y_test})
result.plot(kind='bar', figsize=(18,9))
```
