Since in this data we have got the range in the cats column so i divided it with the minimmum and the maximmum range of the children who might like that story . If there is any value like 0+ we take the minimmum as 0 and the maximmum as 15 because that's the max age of children ( In respect to this data) ....

```python
# Let's sort the data some bit ....
arr=data['cats']
minn=[]
ma=[]
for i in arr:
    x=(i.replace('Age ',''))
    if '-' in x:
        mi,mx=x.split('-')
        mi,mx=int(mi),int(mx)
        minn.append(mi)
        ma.append(int(mx))
    else:
        if x.split('+')[0].isdigit():
            minn.append(int(x.split('+')[0]))
        else:
            minn.append(0)
        ma.append(15)
data['Minimum Age']=minn
data['Maximmum Age']=ma
    
    
    
```

# Now we can make a function to find the specific words you can put in your story for particular range of children to like it

## Removing the stopwords and punctuations from the data using the ntlk and some python :)

```python
stopset = set(stopwords.words('english'))
```

```python
# Function to remove the puctuation marks from my string
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''
def remove_punctuation(my_str):
    no_punct = ""
    for char in my_str:
       if char not in punctuations:
           no_punct = no_punct + char

    # return the unpunctuated string
    return no_punct
```

## Finding out the top words likely to be there in the best of the stories that are there 

```python
def find_words(words,number=10,mi=2,ma=9):
    df=data[(data['Minimum Age']==mi) & (data['Maximmum Age']==ma)]
    d=df['desc'].values
    dic={}
    sentences=[]
    for i in d:
        i=remove_punctuation(i)
        tokens=word_tokenize(str(i))
        tokens = [w for w in tokens if not w in stopset]
        for j in tokens:
            if j not in dic:
                dic[j]=1
            else:
                dic[j]+=1
    dic={k: v for k, v in sorted(dic.items(), key=lambda item: item[1],reverse=True)}
    return dict(itertools.islice(dic.items(), number)) 
    
    
        
            
    
    
```

## Asking the user for the range of the book he might like the words to be from .... and some visualization of those words ....
