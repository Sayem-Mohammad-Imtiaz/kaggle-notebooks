####  This will show the process of filtering a dataframe to display one team
####  & create a bar plot that shows the maximum compensation per position

```python
# parse league list to only NY RedBulls
NYRedBulls_df = mls_salaries_df.set_index("club").loc["NYRB",:]
NYRedBulls_df
```

```python
# reset index so idxmax() can be used to extract max values row
# note after groupby can't reset index
NYRedBulls_df.reset_index(inplace=True)
NYRedBulls_df
```

```python
# group new dataframe by position & get max value (compensation) index row number
grouped_NYRB = NYRedBulls_df.groupby(by="position")['guaranteed_compensation'].idxmax()
```

```python
# reset index so idxmax() can be used to extract max values row
grouped_NYRB
```

```python
NYRB_summary_df = NYRedBulls_df.iloc[grouped_NYRB].sort_values(by=['guaranteed_compensation'])
```

```python
NYRB_summary_df
```

```python
NYRB_summary_df.plot(x="position", y="guaranteed_compensation", ylabel="compensation", kind="bar", title="NYRB")
```

```python
# parse league list to only show club of interest
print("2017 roster and associated salaries")
some_team_df = mls_salaries_df.set_index("club").loc[team,:]
some_team_df
```

```python
# reset index so idxmax() can be used to extract max values row
# note after groupby can't reset index
some_team_df.reset_index(inplace=True)
some_team_df
```

```python
# group new dataframe by position & get max value (compensation) index row number
grouped_some_team = some_team_df.groupby(by="position")['guaranteed_compensation'].idxmax()
# reset index so idxmax() can be used to extract max values row
grouped_some_team
```

```python
team_max_df = some_team_df.iloc[grouped_some_team].sort_values(by=['guaranteed_compensation'])
team_max_df
```

```python
team_max_df.plot(x="position", y="guaranteed_compensation", ylabel="compensation", kind="bar", title=team)
```
