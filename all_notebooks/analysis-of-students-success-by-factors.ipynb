## Numbers and Distrubitions Of Students In Groups

```python
group_list = list(students.race.unique())
numbers = list()

for g in group_list:
    size = students[students.race==g].shape[0]
    numbers.append(size)
```

```python
groupPie = dict(
    values=numbers,
    labels=group_list,
    type="pie",
    domain=dict(x= [0,0.5]),
    text = group_list,
    marker = dict(line=dict(width=2,color="#444")),
    hoverinfo = "value",
    textinfo="percent+text",
    hole= .2
    )
groupFig = go.Figure(data=groupPie)
groupFig.update_layout(
    modebar=dict(activecolor="#111"),
    legend=dict(orientation="v",borderwidth=1,bordercolor="#444",x=0.75),
    paper_bgcolor="#d8d8d8",
    title=dict(text="Numbers and Distrubitions Of Students In Groups",x=0.05,font=dict(size=20)),
    hoverlabel=dict(bordercolor="#444",font=dict(color="#fff",size=14))
)
iplot(groupFig)
```

## Mean Of Exams By Groups

```python
groups = students.groupby("race")
means = groups.mean()
means.columns.names = ["mean"]
sorting_groups=groups.race.head(1).sort_values().unique()
means
```

<a id="a"></a>

```python
math = dict(
    type="bar",
    x=means.math_score.index,
    y=means.math_score.values,
    name= "Math Score",
    marker=dict(line=dict(width=2,color="#111")),
    hoverinfo="x+name+y"
)
reading = dict(
    type="bar",
    x=means.reading_score.index,
    y=means.reading_score.values,
    name= "Reading Score",
    marker=dict(line=dict(width=2,color="#111")),
    hoverinfo="x+name+y"
)
writing = dict(
    type="bar",
    x=means.writing_score.index,
    y=means.writing_score.values,
    name= "Writing Score",
    marker=dict(line=dict(width=2,color="#111")),
    hoverinfo="x+name+y"
)
scoresFig=go.Figure(data=(math,reading,writing),layout=dict(barmode="group"))
scoresFig.update_layout(
    modebar=dict(activecolor="#505050"),
    plot_bgcolor="#505050",
    legend = dict(bordercolor="#505050",borderwidth=1,orientation="h"),
    hoverlabel=dict(bordercolor="#505050",font=dict(color="#fff")),
    title=dict(text="Mean Of Exams By Groups",font=dict(size=20)),
    clickmode="event",
    hovermode="x"
)
scoresFig.update_xaxes(
    ticks="outside",
    ticklen=3,
    tickwidth=3,
    tickcolor="black",
    showticklabels=True,
    tickangle=0,
    tickfont=dict(color="black",size=13),
    showgrid=False,
    zeroline=False)
scoresFig.update_yaxes(
    zeroline=False,
    showgrid=True,
    linewidth=5,
    linecolor="#505050",
    mirror=True,
    title_text="Scores",
    gridwidth=1.25,
    gridcolor="#fff",
    range=(50,75),
    tickfont={"color":"#505050","size":13}
    )
scoresFig.show() # or iplot(scoresFig)
```

## Percentages of Lunch Values by Groups

```python
percents= list()

frame = groups.lunch.value_counts().to_frame()
for i in sorting_groups:
    size = frame.loc[i].sum().iloc[0]
    percent = list(percents.append(j) for j in list(frame.loc[i].lunch.iloc[v]/size*100 for v in range(2)))
    
percentlunch = groups.lunch.value_counts().to_frame()
percentlunch["percents"] = percents
func = lambda x: float("{:.2f}".format(x))
percentlunch["percents"] = percentlunch["percents"].apply(func)
percentlunch.unstack(level=0).percents
```

```python
lunchbar1 = dict(x = percentlunch.unstack().index,
                 y = percentlunch.unstack().percents.T.iloc[1].values,
                 type="bar",
                 name= "Standart",
                 marker = dict(color="#8B0B53",line=dict(color="black",width=2)))
lunchbar2= dict(x = percentlunch.unstack().index,
                 y = percentlunch.unstack().percents.T.iloc[0].values,
                 type="bar",
                 name= "Free/Reduced",
               marker = dict(color="#F3BA13",line=dict(color="black",width=2)))

lunchFig = go.Figure(data = (lunchbar2,lunchbar1),layout=dict(barmode="relative"))
lunchFig.update_layout(
    plot_bgcolor = "#143274",
    hovermode="x",
    legend=dict(orientation="h",bgcolor="#143274",font=dict(color="#d8d8d8"),borderwidth=2,bordercolor="black"),
    hoverlabel=dict(bordercolor="black"),
    title=dict(text="Percentages of Lunch Values by Groups",font=dict(size=20)),
    modebar=dict(activecolor="#143274")
)
lunchFig.update_yaxes(
    range=(25,105),
    tickfont=dict(color="black",size=12),
    title=dict(text="%",font=dict(size=14,color="black")),
    gridcolor="#d8d8d8",
    linewidth=2,
    linecolor="black",
    mirror=True)
lunchFig.update_xaxes(
    zeroline=False,
    tickfont=dict(color="black",size=12),
    title=dict(text="Groups",font=dict(size=14,color="black")),
    linewidth=2,
    linecolor="black",
    mirror=True)
lunchFig.show()
```

<font style="font-size:16px" >Percent of standard lunch type is the least in group A but percent of standard lunch type is the highest in group E. If we look <a style="font-weight:600;" href="#a">these results</a>, we can say ratio of success is linearly proportional with standard lunch type percent.</font>

## Percentages of Genders by Groups

```python
genderPer = list()
for g in sorting_groups:
    serie = students.set_index("race").loc[g].gender.value_counts()
    size = serie.sum()
    for i in range(serie.size):
        per = serie.iloc[i]/size * 100
        genderPer.append(float("{:.2f}".format(per)))
gender_percent = groups.gender.value_counts().to_frame()
gender_percent["gender_percent"] = genderPer
gender_percent.drop("gender",axis=1,inplace=True)
graphgen = gender_percent.unstack().gender_percent
graphgen
```

```python
genderFig = go.Figure(data = (dict(x = graphgen.index,
                                y =graphgen.male,
                                type="bar",name="male",
                                text = "%",
                                hoverinfo="y+text+name",
                                marker=dict(color="#206C64",line=dict(color="black",width=2))
                                ),
                            
                              dict(x = graphgen.index,
                                y =graphgen.female,
                                type="bar",name="female",
                                text = "%",
                                hoverinfo="y+text+name",
                                marker=dict(color="#AE6733",line=dict(color="black",width=2)))
                            )
                     )
genderFig.update_layout(
    title=dict(text="Percentages of Genders by Groups",font=dict(size=20)),
    plot_bgcolor="#A9323B",
    hoverlabel=dict(font=dict(color="#fff"),bordercolor="black"),
    hovermode="x",
    modebar=dict(activecolor="#A9323B"),
    barmode="group",
    legend=dict(orientation="h",y=1.1,x=0.84)
)
genderFig.update_yaxes(
    zeroline=False,
    gridcolor="gray",
    gridwidth=1,
    linecolor="black",
    linewidth=3,
    mirror=True,
    title_text="%",
    tickfont=dict(size=12,color="gray"),
    titlefont=dict(size=14,color="black")
)
genderFig.update_xaxes(
    linecolor="black",
    linewidth=3,
    mirror=True,
    title=dict(text="Groups",font=dict(size=14,color="black")),
    tickfont=dict(size=12,color="gray")
)
genderFig.show()
```

<font style="font-size:16px">Although it is not exactly like that, we can say that the <a style="font-weight:600;" href="#a">success ranking</a> is higher in groups where gender percentages are close to each other.</font>

<a id="2f"></a>
## Percentages of Course Completions by Groups

```python
none,completed=list(),list()
for i in sorting_groups:
    percnone = groups.course.value_counts().to_frame().loc[i].course.iloc[0] / groups.course.value_counts().to_frame().loc[i].course.sum() * 100
    none.append(percnone)
    perccomp = groups.course.value_counts().to_frame().loc[i].course.iloc[1] / groups.course.value_counts().to_frame().loc[i].course.sum() * 100
    completed.append(perccomp)
```

```python
nonebar = dict(
    x = sorting_groups,
    y = none,
    type="bar",
    marker = dict(color="#B13E00",line=dict(width=2,color="black")),
    name="None",
    text = "%",
    hoverinfo = "y+text+name")
completedbar = dict(
    x = sorting_groups,
    y = completed,
    type="bar",
    marker = dict(color="#B18400",line=dict(width=2,color="black")),
    name="Completed",
    text = "%",
    hoverinfo = "y+text+name")

courseFig = go.Figure(data=(nonebar,completedbar))
courseFig.update_yaxes(
    linewidth=3,
    linecolor="black",
    mirror=True,
    range=(25,80),
    gridcolor="black",
    title=dict(text="%",font=dict(color="black",size=14)),
    tickfont=dict(color="black",size=12)
)
courseFig.update_xaxes(
    linewidth=3,
    linecolor="black",
    mirror=True,
    tickfont=dict(color="black",size=12))
courseFig.update_layout(
    plot_bgcolor="#890059",
    modebar=dict(activecolor="#890059"),
    hovermode="x",
    hoverlabel=dict(bordercolor="black",font=dict(color="#fff")),
    title=dict(text="Percentages of Course Completion by Groups"),font=dict(size=16),
    legend=dict(bordercolor="black",borderwidth=2,orientation="h",bgcolor="#890059",font=dict(color="#000",size=12)))
courseFig.show()
```

<font style="font-size:16px">Apart from the most successful group, we can see that the course completion rate does not affect success much.</font>

<a id="2g"></a>
## Percentages of Parental Educations by Groups

```python
parental,parentalval = list(),list()

for i in sorting_groups:
    val = groups.parental_education.value_counts().loc[i]
    size = val.sum()
    for j in range(val.index.size): 
        parentalval.append(val.iloc[j]/size*100)
        parental.append(val.index[j])
        
framelist=list()
fr = pd.DataFrame(parental,parentalval)
for v in fr[0].unique():
    frame = fr.groupby(0).get_group(v)
    frame["groups"] = sorting_groups
    framelist.append(frame)
for i in framelist:
    i.reset_index(inplace=True)
```

```python
tracelist=list()
colorlist = ["#97042C","#A78205","#1A1374","#409304","#A32812","#0B6758"]
for number,name,color in zip(range(len(framelist)),fr[0].unique(),colorlist):
    trace = dict(
        x = framelist[number]["groups"],
        y = framelist[number]["index"],
        type="bar",
        hoverinfo="y+text",
        name = name,
        text = framelist[number][0],
        marker = dict(color=color,line=dict(width=2,color="black")))
    tracelist.append(trace)

figPar = go.Figure(data=tracelist,layout=dict(barmode="group"))
figPar.update_layout(
    plot_bgcolor = "#2B1319",
    modebar=dict(activecolor="#2B1319"),
    legend=dict(orientation="h",bgcolor="#2B1319",font=dict(color="#fff"),borderwidth=2,bordercolor="black"),
    title=dict(text="Percentages of Parental Educations by Groups",font=dict(size=20)),
    clickmode="event+select"
)
figPar.update_yaxes(
    tickfont=dict(color="black",size=12),
    title=dict(text="%",font=dict(size=14,color="black")),
    linewidth=3,
    linecolor="black",
    mirror=True,
    zeroline=False)
figPar.update_xaxes(
    tickfont=dict(color="black",size=12),
    linewidth=3,
    linecolor="black",
    mirror=True
)
figPar.show()
```

<font style="font-size:16px"> There doesn't seem a significant relationship between <a href="#success" style="font-weight:600">success ranking</a> and parental educations. Apart of group D, just some high school percentages in groups are opposite proportional with <a href="#success" style="font-weight:600">success ranking</a>. </font>
