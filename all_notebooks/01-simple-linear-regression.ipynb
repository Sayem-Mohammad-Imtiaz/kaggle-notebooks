# SIMPLE LINEAR REGRESSION
    
  It is Statistical method that allows us to Predict a quantitative response Y on the basis of single
  
  predictor variable x.
                    
Ex:Performance of a student in exam depends no. of hours studied, no. of classes attended, no. of

problems solved etc.

# BASIC ASSUMPTIONS?

There are four assumptions associated with a Linear Regression Model.

1. Linearity: The relationship between X and Y is linear.

2. Homoscedasticity: The variance of residual is the same for any value of X.

3. Independence: Observations are independent of each other.

4. Normality: For any fixed value of X,Y is normally distributed.

If the data is not distribUted normally, we can make into normally distributed by using 

transformations like log transformation, box plot transformation,exponetional transformation, 

power log transformation.


# PROS

1. It Performs Exceptionally well for Linear Seperable Data.

2. Easy to Implement and train the model.

3. It can handle overfitting using dimensionality reduction techniques and cross validation.
   
   
# CONS

1. Sometimes lot of Feature Engineering is required.

2. If the independent features are considered it may affect performance.

3. It is often quite prone to noise and overfitting.




# WHETHER FEATURE SCALING IS REQUIRED?

Yes,because Feature Scaling helps to reach Global Minima Quickly.

# Impact of Missing Values

* It is Senstive to Missing Values.

* Some times we need to do lot of Feature Engineering, if there are lot of missing values, we will use lot of strategies and replace the value, and if the strategy doesn't work well it will definitely be sensitive to missing values.

* Linear Regression doesn't handle missing values, we have to handle missing values by doing Feature Engineering.


# IMPACT OF OUTLIERS

* Linear Regression needs the relationship between the independent and dependent variables to be linear.It is also important to check for outliers, since linear regression is sensitive to outlier effects.

# IMPORTING LIBRARIES

```python
import numpy as np # linear algebra
import matplotlib.pyplot as plt # Plotting
import pandas as pd # data processing, CSV file I/O
```

# READING AND UNDERSTANDING DATA

```python
df = pd.read_csv('../input/random-salary-data-of-employes-age-wise/Salary_Data.csv')
```

# To display first five columns and rows

```python
df.head()
```

# TO DISPLAY HOW MANY ROWS AND COLUMNS WERE PRESENT

```python
df.shape
```

* # CHECKING NULL VALUES

```python
df.isnull().sum()
```

# SEPARATING X AND Y VALUES

```python
X = df.iloc[:,:-1].values 
Y = df.iloc[:,1].values
```

```python
X
```

```python
Y
```

# SPLITTING THE DATA SET INTO TRAINING SET AND TEST SET

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 0)
```

```python
X_train
```

```python
Y_train
```

# MODEL BUILDING

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#bias(b)
model.intercept_
```

```python
#slope/weight (w)
model.coef_
```

# PREDICTIONS

```python
Y_pred = model.predict(X_test)
```

```python
Y_pred
```

# MODEL EVALUATION

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(Y_test,Y_pred)
```

```python
score
```

# VISUALISING THE TRAINING SET RESULTS

```python
plt.scatter(X_train,Y_train,color = 'red')
plt.plot(X_train,model.predict(X_train),color = 'blue')
plt.title('Salary vs Year of experience(training set)')
plt.xlabel('Years of exp')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X_test,Y_test,color = 'red')
plt.plot(X_train,model.predict(X_train),color = 'blue')# no need to give the X_test,y_pred values
plt.title('Salary vs Year of experience(testing set)')
plt.xlabel('Years of exp')
plt.ylabel('Salary')
plt.show()
```
