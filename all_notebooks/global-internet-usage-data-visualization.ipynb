```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
import plotly.express as px
from pandas_profiling import ProfileReport
```

```python
df = pd.read_csv("/kaggle/input/gapminder-internet/gapminder_internet.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
report = ProfileReport(df)
report
```

```python
df = df.dropna()
```

```python
plt.figure(figsize = [10,7])
sns.set(font_scale = 1.5)
df1 = df.sort_values(by = ['internetuserate'],ascending = False).reset_index().head(10)              
sns.barplot(x = df1['internetuserate'], y = df1['country'],palette = 'cividis')
```

```python
plt.figure(figsize = [10,7])
sns.set(font_scale = 1.5)
df1 = df.sort_values(by = ['incomeperperson'],ascending = False).reset_index().head(10)              
sns.barplot(x = df1['incomeperperson'], y = df1['country'],palette = 'BuPu')
```

```python
plt.figure(figsize = [10,7])
sns.set(font_scale = 1.5)
df1 = df.sort_values(by = ['urbanrate'],ascending = False).reset_index().head(10)              
sns.barplot(x = df1['urbanrate'], y = df1['country'],palette = 'crest')
```

```python
df1 = df.groupby('country').sum().sort_values('internetuserate',ascending = False).reset_index().head(10)
fig = px.scatter(df1,
                 x = 'internetuserate',
                 y = 'incomeperperson',
                 color = 'country',
                 color_discrete_sequence = px.colors.sequential.Rainbow)

fig.update_traces(marker = dict(size = 25,
                                line = dict(color = 'rgb(0, 0, 0)',width = 1.5)),
                  selector = dict(mode = 'markers'))
fig.show()
```

```python
df1 = df.groupby('country').sum().sort_values('internetuserate',ascending = False).reset_index().head(10)
fig = px.scatter(df1,
                 x = 'internetuserate',
                 y = 'urbanrate',
                 color = 'country',
                 color_discrete_sequence = px.colors.sequential.Jet)

fig.update_traces(marker = dict(size = 25,
                                line = dict(color = 'rgb(0, 0, 0)',width = 1.5)),
                  selector = dict(mode = 'markers'))
fig.show()
```

```python
fig = go.Figure(data = go.Choropleth(locations = df['country'],
                                     z = df['incomeperperson'],
                                     locationmode = 'country names',
                                     colorscale = 'twilight',
                                     autocolorscale = False,
                                     reversescale = True,
                                     marker_line_color = 'LightGray',
                                     marker_line_width = 0.3,
                                     colorbar_title = 'Income/per person'))

fig.update_traces(colorbar_title_font = dict(family = 'Droid Sans',
                                             size = 14,
                                             color = 'DarkBlue'))
fig.update_layout(geo = dict(showframe = False,
                             showcoastlines = False,
                             projection_type= 'equirectangular'))
fig.show()  
```

```python
df1 = df.sort_values(by = ['internetuserate'],ascending = False).reset_index().head(50)
fig = px.choropleth(df1,
                    locations = 'country',
                    locationmode = 'country names',
                    color = 'internetuserate',
                    animation_frame = 'country',
                    color_continuous_scale = 'plasma')

fig.update_layout(geo = dict(showframe = False,
                             showcoastlines = False,
                             projection_type = 'natural earth'))

fig.show()
```
