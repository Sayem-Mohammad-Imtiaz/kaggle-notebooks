```python
# function that returns great of two numbers

def func_greater(a,b):
    
    if (a>b):
        return ('first integer ' + str(a))
    elif (a==b):
        return 'equal'
    else:
        return ('second integer ' + str(b))
```

```python
import matplotlib.pyplot as plt
```

```python
_, ax = plt.subplots(1, 1, figsize=(18, 9))

df_subset = df_pgen1[df_pgen1['DATE']=='2020-05-21']


ax.plot(df_subset.DATE_TIME,
        df_subset.DC_POWER/10,
        marker = 'o',
        linestyle='',
        label='DC_POWER'
       )

ax.plot(df_subset.DATE_TIME,
        df_subset.AC_POWER,
        marker = 'o',
        linestyle='',
        label='AC_POWER'
       )

ax.grid()
ax.margins(0.05)
ax.legend()
plt.title('DC Power and AC Power over 34 Days')
plt.xlabel('Date and Time')
plt.ylabel('Power')
plt.show()
```
