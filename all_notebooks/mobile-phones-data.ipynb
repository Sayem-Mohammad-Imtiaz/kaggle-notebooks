```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns, plotly.express as px, plotly.figure_factory as ff
import warnings
warnings.filterwarnings('ignore')
```

```python
mobiles = pd.read_csv('../input/ukrainian-market-mobile-phones-data/phones_data.csv', index_col=0)
```

```python
# Fix release date kind.
mobiles['release_date'] = pd.to_datetime(mobiles['release_date'])
```

```python
print(f"This datasets covers mobile phones sold in Ukraine between {mobiles['release_date'].min()} and {mobiles['release_date'].max()}")
print(f"This datasets has {mobiles.shape[0]} columns and {mobiles.shape[1]} rows.")
print(f"This datasets has {mobiles.isna().sum().sum()} missing values")
print(f"Missing values are distributed as such: {mobiles.isna().sum()}")
```

I see that 

```python
#Fixing numericals NaN
numericals = mobiles.dtypes[(mobiles.dtypes!='O') & (mobiles.dtypes!='<M8[ns]')].index.tolist()
mobiles[numericals] = mobiles[numericals].fillna(mobiles[numericals].mean())

#Fixing categoricals NaN
categoricals = mobiles.dtypes[mobiles.dtypes == 'O'].index.tolist()
mobiles[categoricals] = mobiles[categoricals].fillna('unknown')
```

```python
#Fixing Dtypes
mobiles[numericals] = mobiles[numericals].astype(int)
mobiles[categoricals] = mobiles[categoricals].astype('category')
mobiles.info()
```

```python
# Converting currency. From UAH to EUR
ua_prices = ['best_price', 'lowest_price','highest_price']
eu_prices = ['eu_best_price', 'eu_lowest_price','eu_highest_price']

mobiles[eu_prices] = mobiles[ua_prices] / 32
```

# When most of the phones were released?

```python
viz_data = mobiles.copy(True)
viz_data['release_date'].value_counts(normalize=True).sort_values(ascending=False)
```

```python
viz_data['release_date'].value_counts()
viz_data['release_date'].hist()
plt.xlabel('Year')
plt.ylabel('Number of releases')
plt.show()
```

# Which brands are popular in Ucraine? 


```python
# Which mobile brands are the most popular in Ucraine?
br_data = mobiles.copy(True)
br_data = mobiles.groupby('brand_name').count()
br_data = br_data.sort_values('model_name', ascending=False)
```

```python
px.bar(x = br_data.index, y = br_data.model_name, color_discrete_sequence=['orange'], 
       labels={'x': 'Brand', 'y': 'Phones amount'})
```
