```python
import pandas as pd
import plotly.graph_objects as go
import numpy as np
import datetime
from plotly.subplots import make_subplots

from plotly.offline import init_notebook_mode, iplot
init_notebook_mode(connected=True)
```

```python
bvc_data = pd.read_csv("../input/colombia-stock-market-data/BVC.csv")
eco_data = pd.read_csv("../input/colombia-stock-market-data/ECOPETROL.csv")
bvc_data.head()
```

```python
fig = go.Figure(data=go.Scatter(x=bvc_data.fecha, y=bvc_data["Precio Cierre"]))

fig.update_layout(title='Precio de la acción BVC',
                  xaxis_title='Fecha',
                   yaxis_title='Precio (En $COP)')

fig.show()
```

```python
fig = go.Figure(data=go.Scatter(x=eco_data.fecha, y=eco_data["Precio Cierre"]))

fig.update_layout(title='Precio de la acción ECOPETROL',
                  xaxis_title='Fecha',
                   yaxis_title='Precio (En $COP)')

fig.show()
```

```python
fig = go.Figure(data=go.Histogram(y=eco_data["Precio Cierre"]))

fig.update_layout(title='Histograma de precios - acción ECOPETROL',
                  xaxis_title='Fecha',
                   yaxis_title='Precio (En $COP)')

fig.show()
```
