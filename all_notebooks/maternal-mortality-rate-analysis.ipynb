```python
df=pd.read_csv('../input/maternal-mortality-ratiommr/maternal-mortality.csv')
df.head()
```

```python
gk=df.groupby('Entity')
newdf=gk.first()
```

```python
newdf.info()
```

```python
newdf.head()
```

```python
data=newdf.swapaxes("index", "columns")
```

**TOP 10 COUNTRIES WITH HIGHEST MATERNAL MORTALITY RATE**

```python
datadict={}
for i,j in enumerate(data):
    #print(i,j)
    total=data.iloc[1:,i].sum()
    datadict[j]=total

s=pd.Series(datadict)
fr=s.to_frame()
top_10countries=fr.sort_values(0,ascending=False)[:10]
```

```python
sns.set_theme()
plt.figure(dpi=100,figsize=(10,6))
plt.style.use('ggplot')
plt.title('TOP 10 COUNTRIES WITH HIGHEST MATERNAL MORTALITY RATE',color='#F4C430')
plt.bar(top_10countries.index,top_10countries[0])
plt.xticks(rotation=45)
for index, value in enumerate(top_10countries[0]):
    plt.text(index,value+300, str(str(int(value/1000))+'k'))
plt.show()
```

**Insight-9 out of top 10 countries with highest maternal mortality rate are located in Africa**

**TOP 10 COUNTRIES WITH LOWEST MATERNAL MORTALITY RATE**

```python
countries_with_least_mmr=fr.sort_values(0)[:10]
sns.set_theme()
plt.figure(dpi=100,figsize=(10,6))
plt.style.use('ggplot')
plt.title('TOP 10 COUNTRIES WITH LOWEST MATERNAL MORTALITY RATE',color='#F4C430')
plt.bar(countries_with_least_mmr.index,countries_with_least_mmr[0])
plt.xticks(rotation=45)
for index, value in enumerate(countries_with_least_mmr[0]):
    plt.text(index,value+1, str(value))
plt.show()
```

**Insight-9 out of 10 countries having lowest maternal mortality rate are in Europe.**
