2. Escribir un programa que reciba una cadena de caracteres y devuelva un diccionario con cada palabra que contiene y su frecuencia. Escribir otra función que reciba el diccionario generado con la función anterior y devuelva una tupla con la palabra más repetida y su frecuencia.


````python
def fibonacci(iterations):
    previousValue = 0
    nextValue=1
    result="0"
    for item in range(1, iterations):
        sumFibonacci = previousValue + nextValue
        previousValue = nextValue
        nextValue = sumFibonacci
        result = (f'{result},{str(previousValue)}')
    return result
print(fibonacci(9))
````

````python
def most_frequent(items):
    dict = {}
    count, result = 0, ''
    for item in reversed(items):
        dict[item] = dict.get(item, 0) + 1
        if dict[item] >= count :
            count, result = dict[item], item
    return(result)
  
items = ['Dog','Cat','Wolf','Dog']
print(most_frequent(items))
````
