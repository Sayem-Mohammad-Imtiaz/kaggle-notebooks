Check for Null values

```python
data.isnull().sum()
```

```python
data.info()
```

```python
data['Item_Weight'].fillna(data['Item_Weight'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Outlet_Size'].mode()
```

```python
data['Outlet_Size'].fillna("Medium",inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Outlet_Type'].value_counts()
```

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
data["Outlet_Type"] = ord_enc.fit_transform(data[["Outlet_Type"]])
data['Outlet_Location_Type'] =ord_enc.fit_transform(data[["Outlet_Location_Type"]])
data['Outlet_Size'] =ord_enc.fit_transform(data[["Outlet_Size"]])
data['Item_Fat_Content'] =ord_enc.fit_transform(data[["Item_Fat_Content"]])
data['Item_Type'] =ord_enc.fit_transform(data[["Item_Type"]])
```

```python
data.head()
```

```python
x = data.drop(['Item_Identifier','Outlet_Identifier','Item_Outlet_Sales'],axis =1)
```

```python
x.shape
```

```python
target = data['Item_Outlet_Sales']
```

```python
tst.head()
```

```python
tst.isnull().sum()
```

```python
tst['Item_Weight'].fillna(tst['Item_Weight'].mean(),inplace=True)
tst['Outlet_Size'].fillna("Medium",inplace=True)
```

```python
tst.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,target,test_size =0.2,random_state=55)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
regr.fit(X_train, y_train)
```

```python
y_linreg = regr.predict(X_test)
```

```python
y_linreg
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

MSE=mean_squared_error(y_test,y_linreg)
MAE=mean_absolute_error(y_test,y_linreg)
r2=r2_score(y_test,y_linreg)
RMSE = np.sqrt(MSE)
print("R squared value: ", r2)
print("Root Mean Squared Error : ", RMSE)
print("Mean Absolute Error : ", MAE)
```

```python
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor(n_estimators=100)
reg.fit(X_train,y_train)
```

```python
reg.predict(X_test)
```

```python
rfr_reg=RandomForestRegressor(n_estimators=200)
rfr_reg.fit(X_train,y_train)

y_rfreg = rfr_reg.predict(X_test)
```

```python
MSE=mean_squared_error(y_test,y_rfreg)
MAE=mean_absolute_error(y_test,y_rfreg)
r2=r2_score(y_test,y_rfreg)
RMSE = np.sqrt(MSE)
print("R squared value: ", r2)
print("Root Mean Squared Error : ", RMSE)
print("Mean Absolute Error : ", MAE)
```
