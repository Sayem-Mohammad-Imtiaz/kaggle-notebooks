```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('../input/coursera-course-dataset/coursea_data.csv')
df.head()
```

```python
plt.style.use('ggplot')
```

```python
fig = plt.figure(figsize=(25,10))
y = list(df['course_organization'].value_counts()[0:10])
x = list(df['course_organization'].value_counts().index[0:10])
plt.bar(x=x,height=y,width=.8)
plt.title('Top 10 course organization',fontdict={'fontsize':36})
```

```python
plt.figure(figsize=(10,5))
nums = df['course_difficulty'].value_counts()
plt.axis('off')
plt.text(0.25,0.8,'Course difficulty',fontsize=30,fontweight='bold',color='#fddc5c')
plt.text(0.2,0.55,nums['Beginner'],fontsize=25,fontweight='bold',color='#1b1407')
plt.text(0.19,0.49,'Beginner',fontsize=12,fontweight='bold',color='#065b3c')
plt.text(0.7,0.55,nums['Intermediate'],fontsize=25,fontweight='bold',color='#1b1407')
plt.text(0.675,0.49,'Intermediate',fontsize=12,fontweight='bold',color='#065b3c')
plt.text(0.19,0.3,nums['Mixed'],fontsize=25,fontweight='bold',color='#1b1407')
plt.text(0.2,.24,'Mixed',fontsize=12,fontweight='bold',color='#065b3c')
plt.text(0.715,0.3,nums['Advanced'],fontsize=25,fontweight='bold',color='#1b1407')
plt.text(0.69,.24,'Advanced',fontsize=12,fontweight='bold',color='#065b3c')
```

```python
for i in range(len(df['course_students_enrolled'])):
  if df['course_students_enrolled'][i].endswith('k'):
    df['course_students_enrolled'][i] = df['course_students_enrolled'][i].replace('k','')
    df['course_students_enrolled'][i] = float(df['course_students_enrolled'][i]) / 10
  elif df['course_students_enrolled'][i].endswith('m'):
    df['course_students_enrolled'][i] = df['course_students_enrolled'][i].replace('m','')
    df['course_students_enrolled'][i] = float(df['course_students_enrolled'][i]) * 100
```

```python
Beginner = df['course_difficulty'] == 'Beginner'
Beginner = df[Beginner]['course_students_enrolled'].sum()
Intermediate = df['course_difficulty'] == 'Intermediate'
Intermediate = df[Intermediate]['course_students_enrolled'].sum()
Mixed = df['course_difficulty'] == 'Mixed'
Mixed = df[Mixed]['course_students_enrolled'].sum()
Advanced = df['course_difficulty'] == 'Advanced'
Advanced = df[Advanced]['course_students_enrolled'].sum()

plt.bar(x=['Beginner','Intermediate','Mixed','Advanced'],
        height=[Beginner,Intermediate,Mixed,Advanced],
        color=['#1e4d5a','#e73f23','#ffc000','#9ac83c'])
```

```python
plt.figure(figsize=(10,4))
nums = df['course_Certificate_type'].value_counts()
plt.axis('off')
plt.text(0.08,0.7,'Course Certificate type',c='#fddc5c',fontsize=36,fontweight='bold')
plt.text(0.08,0.7,'Course Certificate type',c='#fddc5c',fontsize=36,fontweight='bold')
plt.text(0.08,.35,'582',c='#1b1407',fontsize=25,fontweight='bold')
plt.text(0.085,.29,'Course',c='#0cb577',fontsize=12,fontweight='bold')
plt.text(0.4,.35,'297',c='#1b1407',fontsize=25,fontweight='bold')
plt.text(0.35,.29,'SPECIALIZATION',c='#0cb577',fontsize=12,fontweight='bold')
plt.text(0.77,.35,'12',c='#1b1407',fontsize=25,fontweight='bold')
plt.text(0.65,.29,'PROFESSIONAL CERTIFICATE',c='#0cb577',fontsize=10,fontweight='bold')
```
