```python
sess = tf.InteractiveSession()
tf.global_variables_initializer().run()
```

```python
class Data():

    def __init__(self, data, labels):
        self._data = data
        self._labels = labels
        self._index = 0
        self.N = data.shape[0]
        self._labels = labels
        self.D = len(set(self._labels))
        
    def next_batch(self, number):
        end = min(self._index + number, self.N-1)
        ret = self._data[self._index:end, :]
        ret_labels = np.zeros((ret.shape[0], self.D))
        ret_labels[np.arange(ret.shape[0]), self._labels[self._index:end]] = 1
        self._index += number
        if self._index >= self.N:
            self._index = 0
        return ret, ret_labels
    
    def all(self):
        ret_labels = np.zeros((self.N, self.D))
        ret_labels[np.arange(self.N), self._labels] = 1
        ret = self._data
        return ret, ret_labels
    
def labels_to_values(labels):
    distinct_labels = sorted(list(set(labels)))
    label_dict = {distinct_labels[x]:x for x in range(0, len(distinct_labels))}
    values = np.array([label_dict[x] for x in labels])
    return values, label_dict
                                                      
```

```python
values, label_dict = labels_to_values(data.as_matrix()[:, -1])
import numpy.random as random
random.seed(10)
randomize = random.randint(data.shape[0])
print(randomize)
train_data = Data(data.as_matrix()[randomize, 0:-1], 
                  values)
```

```python
for _ in range(5000):
    batch_xs, batch_ys = train_data.all()
    sess.run(train_step, feed_dict={x: batch_xs, y_: batch_ys})
```

```python
correct_prediction = tf.equal(tf.argmax(y, 1), tf.argmax(y_, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
test_x, test_y = train_data.all()
print(sess.run(accuracy, feed_dict={x: test_x,
                                    y_: test_y}))
```
