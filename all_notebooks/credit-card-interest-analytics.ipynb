```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (7,7)

tarr = ['Female', 'Male']
tarr.sort()

outs = [0]*len(tarr)

f = open("/kaggle/input/credit-card-lead-prediction/train.csv", "r")

lines = f.readlines()

for i in range(1, len(lines)):
    line = lines[i].replace("\n", "").split(",")[1]
    
    outs[tarr.index(line)] += 1

plt.title("Female vs Male")
plt.pie(outs, labels=tarr, autopct="%1.2f%%", explode=(0.02, 0.02))
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (15,10)

tarr = ["Yes", "No", "0", "1"]
labels = ["Actived", "Not Actived", "Lead", "Not Lead"]
tarr.sort()

outs = [0]*len(tarr)

f = open("/kaggle/input/credit-card-lead-prediction/train.csv", "r")

lines = f.readlines()

for i in range(1, len(lines)):
    line = lines[i].replace("\n", "").split(",")
    
    outs[tarr.index(line[9])] += 1
    outs[tarr.index(line[10])] += 1
    
plt.xlabel("Labels")
plt.ylabel("Total number")
plt.title("Active and Lead status")
plt.bar(labels, outs, width=0.618, color="darkgrey")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (19,9)

tarr = ['FemaleOther', 'FemaleSalaried', 'FemaleSelf_Employed', 'FemaleEntrepreneur', 'MaleOther', 'MaleSalaried', 'MaleSelf_Employed', 'MaleEntrepreneur']
tarr.sort()

outs = [0, 0, 0, 0]*2

f = open("/kaggle/input/credit-card-lead-prediction/train.csv", "r")

lines = f.readlines()

for i in range(1, len(lines)):
    line = lines[i].replace("\n", "").split(",")
    
    outs[tarr.index(line[1]+line[4])] += 1

plt.xlabel("Number of jobs by gender and employment")
plt.xlabel("Gender and gender of employment")
plt.title("Female vs Male occupation")
plt.bar(tarr, outs, color=("grey", "grey", "grey", "grey", "orange", "orange", "orange", "orange"), width=0.718)
plt.show()
```
