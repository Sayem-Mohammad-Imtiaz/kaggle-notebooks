```python
def remove_dollor(x):
    a =  x[2:] #removes first two chr
    result = ""
    for i in a:
        if i.isdigit() is True:
            result = result + i
    return result #returns only digits (excludes special character)
```

```python
df["hoa"] = pd.to_numeric(df["hoa"].apply(remove_dollor), errors= "ignore")
print(df.hoa.head())
```
