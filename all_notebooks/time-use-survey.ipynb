```python
#if age between these values then group them as below
def groupByAge(row):
    if row['age'] > 0 and row['age'] <= 16:
        val = "Child"
    elif row['age'] >=17 and row['age'] <=30:
        val = "Young Adults"
    elif row['age'] >=31 and row['age'] <=45:
        val = "Middle-aged Adults"
    else:
        val = "Old-aged Adults"
    return val
```

```python
df['Age-Group'] = df.apply(groupByAge, axis=1)
```

```python
#Group activities by code
def groupByActivities(row):
    if row['B3_q5_c7'] >= 111 and row['B3_q5_c7'] <= 119:
        val = "Crop farming, kitchen gardening"
    elif row['B3_q5_c7'] >=121 and row['B3_q5_c7'] <=129:
        val = "Animal husbandry"
    elif row['B3_q5_c7'] >=131 and row['B3_q5_c7'] <=139:
        val = "Fishing, Forestry, Horticulture, Gardening"
    elif row['B3_q5_c7'] >=140 and row['B3_q5_c7'] <=149:
        val = "Collection of fruit, water, plants, storing and hunting"
    elif row['B3_q5_c7'] >=152 and row['B3_q5_c7'] <=159:
        val = "Processing & Storage"
    elif row['B3_q5_c7'] >=161 and row['B3_q5_c7'] <=169:
        val = "Mining, quarrying, digging, cutting"
    elif row['B3_q5_c7'] >=211 and row['B3_q5_c7'] <=219:
        val = "Construction Activities"
    elif row['B3_q5_c7'] >=221 and row['B3_q5_c7'] <=229:
        val = "Manufacturing Activities"
    elif row['B3_q5_c7'] >=311 and row['B3_q5_c7'] <=319:
        val = "Trade and Business"
    elif row['B3_q5_c7'] >=321 and row['B3_q5_c7'] <=329:
        val = "Services"
    elif row['B3_q5_c7'] >=411 and row['B3_q5_c7'] <=491:
        val = "Household Maintenance and Management"
    elif row['B3_q5_c7'] >=511 and row['B3_q5_c7'] <=591:
        val = "Care for Children, sick, disabled for own Household"
    elif row['B3_q5_c7'] >=611 and row['B3_q5_c7'] <=691:
        val = "Community Services"
    elif row['B3_q5_c7'] >=711 and row['B3_q5_c7'] <=791:
        val = "Learning"
    elif row['B3_q5_c7'] >=811 and row['B3_q5_c7'] <=892:
        val = "Social and Cultural Activities"
    else:
        val = "Personal Care and Self-Maintenance"
    return val
```

```python
df['Activities-Group'] = df.apply(groupByActivities, axis=1)
```
