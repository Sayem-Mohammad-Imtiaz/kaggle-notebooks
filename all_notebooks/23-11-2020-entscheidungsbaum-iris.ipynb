```python
%matplotlib inline
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn import metrics
```

# Laden der Daten in ein Pandas DataFrame

```python
df = pd.read_csv("/kaggle/input/iris-flower-dataset/IRIS.csv")
```

# Datenvorverarbeitung

```python
y = df['species']
```

```python
class_names = list(y.unique())
```

```python
class_names
```

```python
X = df.drop(columns='species')
```

```python
feature_names = list(X.columns)
```

```python
feature_names
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=10)
```

# Entscheidungsbaum

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
```

```python
# Oder z.B
# clf = DecisionTreeClassifier(max_depth=2, min_samples_leaf=4)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
print(metrics.classification_report(y_test, y_pred, digits=3))
```

```python
metrics.plot_confusion_matrix(clf, X_test, y_test, cmap='Blues')
plt.grid(False)
plt.show()
```

```python
from yellowbrick.model_selection import ValidationCurve
vc_viz = ValidationCurve(clf, param_name="max_depth", param_range=np.arange(0, 15, 1), cv=10, n_jobs=-1)
vc_viz.fit(X, y)
vc_viz.show()
```

### Entscheidungsbaum ausgeben

```python
feature_names = list(X.columns)
```

```python
class_names = list(y.unique())
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(7,7))
plot_tree(clf, feature_names=feature_names, class_names=class_names, filled = True, rounded=True)
plt.show()
```

# Lernkurve

```python
from yellowbrick.model_selection import LearningCurve
vc_viz = LearningCurve(clf, cv=10, n_jobs=-1)
vc_viz.fit(X, y)
vc_viz.show()
```

# Dummy Classifier

```python
from sklearn.dummy import DummyClassifier
dummy = DummyClassifier()
dummy.fit(X_train, y_train)
y_pred = dummy.predict(X_test)
print(metrics.classification_report(y_test, y_pred, digits=3))
```

```python
from yellowbrick.model_selection import LearningCurve
vc_viz = LearningCurve(dummy, cv=10, n_jobs=-1)
vc_viz.fit(X, y)
vc_viz.show()
```
