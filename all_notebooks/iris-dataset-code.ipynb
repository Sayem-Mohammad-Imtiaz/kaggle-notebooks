```python
import pandas as pd
from sklearn.model_selection import train_test_split
df=pd.read_csv("../input/iris-dataset/My_Iris_Dataset.csv")
```

```python
df=df.drop("Id",axis=1)
x=df.drop("Species",axis=1)
y=df["Species"]
print("x",x.shape)
print("y",y.shape)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=51)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
#sc=StandardScaler()
X_train_sc=sc.fit(X_train)
X_train_sc=sc.transform(X_train)
sc.fit(X_test)
X_test_sc=sc.transform(X_test)
```

```python
X_train_sc=pd.DataFrame(X_train_sc,columns=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
X_test_sc=pd.DataFrame(X_test_sc,columns=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm'])
```
