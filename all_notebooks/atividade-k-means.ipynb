```python

cars = pd.read_csv("../input/carsdata/cars.csv",na_values = ' ' )
cars.head()
```

```python
cars.columns = ['mpg', ' cylinders', 'cubicinches', 'hp', 'weightlbs', 'time-to-60',
       'year', ' brand']
```

```python
cars.describe()
```

```python
cars = cars.dropna()
cars['cubicinches']=cars['cubicinches'].astype(int)
cars['weightlbs'] =cars['weightlbs'].astype(int)
```

```python
X = cars.iloc[:,:7]
X.head()
```

```python
X_array = X.values
```

```python
X_array
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init = 'k-means++', max_iter=300,n_init=10, random_state=0)
    kmeans.fit(X_array)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Cars')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
import sklearn.datasets as dt
dic = dt.load_digits()
dic.keys()
```

```python
dic.data.shape
```

```python
dic.images.shape
```

```python
import matplotlib.pyplot as plt
plt.imshow(dic.images[200])
```

```python
X = dic.data
y = dic.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y)
```

```python
y_train.shape
X_train.shape
```

```python
y_test.shape
y_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7)
knn
model = knn.fit(X_train, y_train)
y_pred = model.predict(X_test)
y_score = model.score(X_test, y_test)
y_pred
y_score
```

```python
import sklearn.metrics
model = str(round(model.score(X_test,y_test) * 100, 2))+"%"
print("O modelo k-NN foi",model)
```
