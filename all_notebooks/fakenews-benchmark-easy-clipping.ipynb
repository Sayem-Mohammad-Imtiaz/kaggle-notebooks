## [fakenews-benchmark](https://www.kaggle.com/kami634/fakenews-benchmark)のアウトプットを利用しています。

```python
import pandas as pd

df = pd.read_csv('../input/fakenewsbenchmark/submission.csv')
```

```python
# 今回isfakeが[0,2]であるはずなのに、その範囲を超えている値があるようです。
df.describe()
```

```python
# fakenews-benchmarkに後処理として[0,2]の範囲で収めて上げます。
df['isfake'] = df['isfake'].clip(0,2)
```

```python
df.describe()
```

```python
df.to_csv('submission.csv',index=False)
```

僅かながら、LBで0.33500 -> 0.33444の改善がみられました!!
