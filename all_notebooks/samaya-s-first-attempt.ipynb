```python
df = pd.read_csv('../input/Chicago_Crimes_2005_to_2007.csv',  names=['ID','Case Number','Date','Block','IUCR','Primary Type','Description','Location Description','Arrest','Domestic','Beat','District','Ward','Community Area','FBI Code','X Coordinate','Y Coordinate','Year','Updated On','Latitude','Longitude','Location'
], error_bad_lines=False, header=None)

df['Month']=[d.split('/')[0] for d in df.Date]

df['Month'] =pd.to_numeric(df['Month'])
df = df.groupby(["Month"], as_index=False).count()
df.Count = df.Date
plt.scatter(df.Month,df.Count, marker='o')
plt.show()
```
