# Importing Required Libraries

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
%matplotlib inline
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
import sklearn.metrics as metrics
import os
```

# Reading the CSV File 

```python
data = pd.read_csv('/kaggle/input/loan-prediction-based-on-customer-behavior/Training Data.csv')
```

```python
data.head() #displaying head of the .csv file
```

# Dropping the Unecessary Columns

```python
data = data.drop("Id", axis = 1)
data = data.drop("CITY", axis=1)
```

```python
data.head() #displaying head of the .csv file
```

# Checking for NULL Values

```python
data.isnull().sum()
```

# Describing the Data

```python
data.describe()
```

# Encoding The Data

```python
en = LabelEncoder()
catCols = ['Married/Single','House_Ownership','Car_Ownership','Profession','STATE']
for cols in catCols:
    data[cols] = en.fit_transform(data[cols])
```

```python
data.head() #displaying head of the .csv file
```

# Splitting Dataset into Train and Test

```python
y = data["Risk_Flag"]
X = data.drop("Risk_Flag",axis = 1)
```

```python
Y=pd.DataFrame(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=0.8, test_size=0.2)
```

# Creating the XGBoost Model

```python
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score, roc_auc_score
# Define the model

    
    # defining the model
clf = XGBClassifier(learning_rate=0.1, 
                    n_estimators=1000, 
                    use_label_encoder=False,
                    random_state=42)
    
clf.fit(X_train, y_train, eval_metric='logloss')
predictions = clf.predict(X_test) 
print("accuracy_score: " + str(accuracy_score(y_test, predictions)))
```

# Calculating the ROC Score of Model

```python
print("ROC AUC score: " + str(roc_auc_score(y_test, predictions)))
```
