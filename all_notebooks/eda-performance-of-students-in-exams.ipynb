### Making another column to calculate the overall percentage

```python
overall_percent=((df["math score"]+df["writing score"]+df["reading score"])/300)*100
df["Overall percentage"]=round(overall_percent,2)
df["Overall percentage"]
```

```python
df.head()
```

### Making another 3 columns for subject wise performance  

```python
def mathprfm(score):
    if score <= 33 :
        return "Poor"
    elif score < 75 and score > 33:
        return "Average"
    elif score >= 75:
        return "Excellent"
df["Math Performance"]=df["math score"].apply(mathprfm)    
```

```python
def readingprfm(score):
    if score <= 33 :
        return "Poor"
    elif score < 75 and score > 33:
        return "Average"
    elif score >= 75:
        return "Excellent"
df["Reading Performance"]=df["reading score"].apply(readingprfm)    
```

```python
def writingprfm(score):
    if score <= 33 :
        return "Poor"
    elif score < 75 and score > 33:
        return "Average"
    elif score >= 75:
        return "Excellent"
df["Writing Performance"]=df["writing score"].apply(writingprfm)    
```

```python
df.head()
```

### To analysis how many students pass or fail in each subject 

```python
def mathstatus(prfm):
    if prfm=="Average" or prfm=="Excellent":
        return 1
    elif prfm=="Poor":
        return 0  
df["Math status"]=df["Math Performance"].apply(mathstatus)
df["Math status"].value_counts()
```

```python
def readingstatus(prfm):
    if prfm=="Average" or prfm=="Excellent":
        return 1
    elif prfm=="Poor":
        return 0   
df["Reading status"]=df["Reading Performance"].apply(readingstatus)
df["Reading status"].value_counts()
```

```python
def writingstatus(prfm):
    if prfm=="Average" or prfm=="Excellent":
        return 1
    elif prfm=="Poor":
        return 0   
df["Writing status"]=df["Reading Performance"].apply(readingstatus)
df["Writing status"].value_counts()
```

## Visualisation under Parental Level of Education wise

```python
countplt, ax = plt.subplots(figsize = (10,5))
ax=sb.countplot(x="parental level of education",data=df,palette="Reds")
```

```python
table = pd.pivot_table(df,
                       index=['gender'],
                       columns = ['parental level of education'],
                       aggfunc={'Overall percentage':np.sum}
                      )
table.plot(kind='bar',figsize = (20,15))
```

From this we come to know that Females whose parenatl level of eduaction is 'some college' got more percentage than who belongs to 'masters degree'. Similarly in Males whose parental level of education is 'associates degree' got more percentage that who belongs to 'masters degree'.

```python
table = pd.pivot_table(df,
                       index=['gender'],
                       columns = ['test preparation course'],
                       aggfunc={'Overall percentage':np.sum}
                      )
table.plot(kind='bar',figsize = (5,5))
```

From this we come to know that in Femlales whose test preperation course was 'none' got more percentage. Also in Males whose test preparation course was 'completed' got more percentage.
