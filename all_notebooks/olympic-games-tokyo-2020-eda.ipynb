# Set up the notebook

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
olympic2020_filepath = "../input/olympic-games-tokyo-2020-medals-per-country/Olympic-Games-Tokyo-2020.csv"

# Read the file into a variable olympic2020_data
# Set index column as "Country"
olympic2020_data = pd.read_csv(olympic2020_filepath, index_col="Country")

# Getting first 10 rows from data frame
olympic2020_data_10 = olympic2020_data.head(10)
```

```python
# Print the first 10 rows of the data, with first 10 row variable
olympic2020_data_10
```

# Line chart

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Line chart showing statics of medals per category
sns.lineplot(data=olympic2020_data_10)

# Setting Y axis value
plt.ylabel("Medals")
```

# Bar chart

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Total medal won by country")

# Preparing bar chart 
sns.barplot(x=olympic2020_data_10.index, y=olympic2020_data_10['Total'])

# Add label for vertical axis
plt.ylabel("Total medal won")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Total 'Gold' medal won by country")

# Preparing bar chart 
sns.barplot(x=olympic2020_data_10.index, y=olympic2020_data_10['Gold'])

# Add label for vertical axis
plt.ylabel("Total medal won")
```

# Heat map chart

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Medal worn by country")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=olympic2020_data_10, annot=True, fmt='g')

# Add label for horizontal axis
plt.xlabel("Country")
```

# Scatter plot

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

sns.scatterplot(x=olympic2020_data_10.index, y=olympic2020_data_10['Total'])
```
