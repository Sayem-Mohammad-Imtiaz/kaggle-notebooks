Basic statistics: mean, standard devation, covariance:
* Mean: $m_x = \frac{1}{n} \sum_{i=1}^{n} x_i$

* Standard deviation: $\sigma_x=\sqrt{\frac{1}{n-1} \sum_{i=1}^n (x_i-m_x)^2}$

* Covariance: $cov_{x,y}=\frac{1}{n-1} \sum_{i=1}^n (x_i-m_x)(y_i-m_y)$

Correlation: $cor_{x,y}=\frac{cov_{x,y}}{\sigma_x \sigma_y}$

ae-y$=a \frac{1}{10^y}$ 

Example: 76e-03$=76 \cdot \frac{1}{10^3}=76 \cdot \frac{1}{1000}=76 \cdot 0.001=0.076$.

The correlation of a variable with itself 1, if the absolute value of the correlation is large, it means that the variables are highly dependent on each other. If the correlation between the variables is close to zero, the variables are almost independent. If the correlation is negative, intuitively, when one variable increases, the other decreases, if it is positive, then vice-versa.

Let us do machine learning (fitting linear regression)

```python
X=df['YearsExperience'] #Predictor
Y=df['Salary']          #variable to be predicted
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
result=train_test_split(X,Y,test_size=0.2)
```

```python
X_train=result[0]  #predictor variables for training
X_test=result[1]   #predictor variables for testing
Y_train=result[2]  #llabels (variables to be predicted) for training
Y_test=result[3]   #labels (variables to be predicted) for testing
#These variable are Series
```
