```python
import plotly.express as px
```

```python
df = pd.read_csv(r'../input/covid19-education-impact-survey/open_one_time_covid_education_impact.csv')
pd.set_option('display.max_column', 40)
df.shape
```

```python
data = df.copy()
```

```python
data.isna().sum()
```

```python
data.info()
```

```python
data.drop('submission_id', axis=1, inplace=True)
```

```python
data.head()
```

```python
values = data['submission_state'].value_counts()
labels = data['submission_state'].value_counts().index

fig = px.bar(data, x=labels, y=values, color_discrete_sequence=['aqua'])
fig.update_layout(title='Submission State', template='plotly_dark', hoverlabel=dict(
    font_size=20
))
fig.update_traces(hovertemplate='%{x} : %{y}')
fig.update_xaxes(title='Submission Status', showline=True, linewidth=1, linecolor='red')
fig.update_yaxes(title='Count', showgrid=False,showline=True, linewidth=1, linecolor='red')
fig.show()
```

```python
data.head()
```

```python
values = data['geography'].value_counts()
labels = data['geography'].value_counts().index

fig = px.pie(data, names=labels, values=values, color_discrete_sequence=['lightgreen', 'lightpink'])
fig.update_layout(title='Geographical Distribution', template='plotly_dark', hoverlabel=dict(
    font_size=20
))
fig.update_traces(hovertemplate='%{label} : %{value}', pull=0.03)
# fig.update_xaxes(title='Submission Status', showline=True, linewidth=1, linecolor='red')
# fig.update_yaxes(title='Count', showgrid=False,showline=True, linewidth=1, linecolor='red')
fig.show()
```

```python
for i in num_cols:
    values = data[i].value_counts()
    labels = data[i].value_counts().index
    
    title_str = str.capitalize(i.replace('_', ' '))                       
    
    fig = px.pie(data, names=labels, values=values, color_discrete_sequence=['lightyellow', 'gray', 'lightpink', 'lightgreen'])
    fig.update_layout(title=f'{title_str}', template='plotly_dark', hoverlabel=dict(
        font_size=20
    ))
    fig.update_traces(hovertemplate='%{label} : %{value}', pull=0.03)
    # fig.update_xaxes(title='Submission Status', showline=True, linewidth=1, linecolor='red')
    # fig.update_yaxes(title='Count', showgrid=False,showline=True, linewidth=1, linecolor='red')
    fig.show()
```
