#### 3.South Korea map for Covid-19 vaccinations 

Professor had taught me how to draw a global map in class, but I want to have a deeper understanding of the situation of South Korea. I want to focus on changes of just one region, but I didn't know how to realize it. So I consult a lot of webpages and materials, download the national boundary data of South Korea, and made a GeoJSON file by myself.And through the method of with open, import my self-made SOUTH_KORA.GEOJSON map file, and use the json_load() function to load the boundary map of south korea, and add geoJSON = KORA_MAP in px.choropleth()function. Finally combine it with the later update_geos() function to achieve partial display.

```python
vaccination_south_korea_df = vaccination_df[vaccination_df['country']== South_Korea_Vaccination]
vaccination_south_korea_df = vaccination_south_korea_df.sort_index()
#print(vaccination_south_korea_df)
with open("../input/south-korea-geojson/south_korea.geojson") as geojson:
    korea_map = json.load(geojson)
    #print(korea_map)

vaccination_fig = px.choropleth(data_frame = vaccination_south_korea_df, geojson = korea_map, locations='country', locationmode = 'country names', 
                                color ='people_fully_vaccinated_per_hundred',hover_name = 'country', 
                                animation_frame = 'date',color_continuous_scale = 'rainbow',range_color = (0.0,5.0),
                                center = {"lat": 37.33, "lon": 126.58})
vaccination_fig.update_geos(fitbounds='locations', visible= True, scope='asia',projection_scale=10000)
vaccination_fig.update_layout(title_text = 'Percentage of fully vaccinated people in South Korea', title_x = 0.4)
vaccination_fig.show()
```

In addition to what professor introduced in class, I carefully read the API documents of Plotly, gained a deeper understanding of the components of Plotly, and added a lot of content according to the content which would be use in my report. I also made the vaccination information for Asian region, especially South Korea, and used the center parameter to set the center point in Seoul(37.33,126.58), because in the previous reports, I  focused on the geographical analysis of South Korea all the time.Since the vaccination rate in South Korea is less than 5%, I adjusted the color range to 0.0-5.0 to get more values and more obvious changes. Finally, when you click the play button of the dynamic graph, first, it will show the asia map because there's no vaccination data in south korea. #### 4.Global map for earthquake worldwide

```python
global_earthquake_df = pd.read_csv('../input/world-earthquake-data/world_earthquake_202105.csv',header=0)
#print(df.columns)
global_earthquake_df.index = pd.to_datetime(global_earthquake_df['time'])
#print(global_earthquake_df.index)
global_earthquake_df['time'] = global_earthquake_df.index.strftime('%Y-%m-%d %H:00:00')
global_earthquake_df = global_earthquake_df.sort_index()
#print(global_earthquake_df['time'])
global_earthquake_figure = px.scatter_geo(global_earthquake_df,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='jet',range_color=(1.0,3.0))
global_earthquake_figure.update_layout(title_text = 'Worldwide Earthquake, May,2021', title_x = 0.4)
global_earthquake_figure.show()
```

```python
WestCoastUS_earthquake_df = pd.read_csv('../input/westcoastu-s-earthquake-data/WestCoastUS202008-09.csv',header=0)
#print(df.columns)
WestCoastUS_earthquake_df.index = pd.to_datetime(WestCoastUS_earthquake_df['time'])
#print(global_earthquake_df.index)
WestCoastUS_earthquake_df['time'] = WestCoastUS_earthquake_df.index.strftime('%Y-%m-%d %H:00:00')
WestCoastUS_earthquake_df = WestCoastUS_earthquake_df.sort_index()
#print(global_earthquake_df['time'])
WestCoastUS_earthquake_figure = px.scatter_geo(WestCoastUS_earthquake_df,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='jet',range_color=(1.0,3.0))
WestCoastUS_earthquake_figure.update_layout(title_text = 'WestCoastUS Earthquake August 2020', title_x = 0.4)
WestCoastUS_earthquake_figure.update_geos(fitbounds="geojson", visible= True, scope='north america')
WestCoastUS_earthquake_figure.show()
```

#### 4. Chile map for earthquake 

```python
chile_earthquake_df = pd.read_csv('../input/chile-earthquake-2021/Chile_earthquake.csv',header=0)
#print(df.columns)
chile_earthquake_df.index = pd.to_datetime(chile_earthquake_df['time'])
#print(global_earthquake_df.index)
chile_earthquake_df['time'] = chile_earthquake_df.index.strftime('%Y-%m-%d %H:00:00')
chile_earthquake_df = chile_earthquake_df.sort_index()
#print(global_earthquake_df['time'])
chile_earthquake_figure = px.scatter_geo(chile_earthquake_df,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='jet',range_color=(1.0,5.0))
chile_earthquake_figure.update_geos(fitbounds="geojson", visible= True, scope='south america', projection_scale=0.1)
chile_earthquake_figure.update_layout(title_text = 'Chile Earthquake in 2021 (2021.1.1-2021.5.31)', title_x = 0.4)
chile_earthquake_figure.show()
```
