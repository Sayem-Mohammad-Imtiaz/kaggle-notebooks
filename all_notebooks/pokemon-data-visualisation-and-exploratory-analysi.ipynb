```python
#first 6 generations only included in the dataset
data = pd.read_csv('../input/pokemonfull/pokemonfull.csv')
```

```python
#subset non legendaries
nonlegendaries = data[data['Legendary'] == False]
#remove megas
nonlegendaries = nonlegendaries[~nonlegendaries.Name.str.contains("Mega")]

top5_attack = nonlegendaries.sort_values(by="Attack", ascending=False).head()
top5_attack

```

```python
#find top 5 attack for each type

for thetype in nonlegendaries["Type 1"].unique():
    print("For " + thetype + " the best 5 Defense pokemons are")
    nonlegendariesoftype = nonlegendaries[(nonlegendaries["Type 1"] == thetype) | (nonlegendaries["Type 2"] == thetype)]
    top5_def = nonlegendariesoftype[['Name','Defense']].sort_values(by="Defense", ascending=False).head()
    print(top5_def)
```

```python
#find top 5 HP for each type

for thetype in nonlegendaries["Type 1"].unique():
    print("For " + thetype + " the best 5 HP pokemons are")
    nonlegendariesoftype = nonlegendaries[(nonlegendaries["Type 1"] == thetype) | (nonlegendaries["Type 2"] == thetype)]
    top5_hp = nonlegendariesoftype[['Name','HP']].sort_values(by="HP", ascending=False).head()
    print(top5_hp)
```

```python
#find top 5 SPEED for each type

for thetype in nonlegendaries["Type 1"].unique():
    print("For " + thetype + " the best 5 Speed pokemons are")
    nonlegendariesoftype = nonlegendaries[(nonlegendaries["Type 1"] == thetype) | (nonlegendaries["Type 2"] == thetype)]
    top5_spd = nonlegendariesoftype[['Name','Speed']].sort_values(by="Speed", ascending=False).head()
    print(top5_spd)
```

```python
#find top 5 Sp. Atk for each type

for thetype in nonlegendaries["Type 1"].unique():
    print("For " + thetype + " the best 5 Sp. Atk pokemons are")
    nonlegendariesoftype = nonlegendaries[(nonlegendaries["Type 1"] == thetype) | (nonlegendaries["Type 2"] == thetype)]
    top5_spat = nonlegendariesoftype[['Name','Sp. Atk']].sort_values(by="Sp. Atk", ascending=False).head()
    print(top5_spat)
```

```python
#find top 5 Sp. def for each type

for thetype in nonlegendaries["Type 1"].unique():
    print("For " + thetype + " the best 5 Sp. def pokemons are")
    nonlegendariesoftype = nonlegendaries[(nonlegendaries["Type 1"] == thetype) | (nonlegendaries["Type 2"] == thetype)]
    top5_spat = nonlegendariesoftype[['Name','Sp. Def']].sort_values(by="Sp. Def", ascending=False).head()
    print(top5_spat)
```

```python
#top 5 of each generation
for i in range(1,7):
    print("For Gen" + str(i) + " the best 5 pokemons are")
    nonlegendariesoftype = nonlegendaries[(nonlegendaries["Generation"] == i)]
    top5tot = nonlegendariesoftype[['Name','Total']].sort_values(by="Total", ascending=False).head()

    print(top5tot)
```
