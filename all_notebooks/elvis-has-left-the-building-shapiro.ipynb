#Let's Shake the Audience. It's always about the Audience. Those that understand it, in

```python
TARGET = 'shake the audience'
```

```python
from scipy.stats import shapiro

SPACE      = 30
WIDTH      = 20
HEIGHT     = 5
MAX_ORD    = 7
MAX_SAMPLE = 5000

def analyze(frame, column):
    # Title
    print(f"{'Column'.rjust(SPACE)} : {column}")
    
    # Missing values
    print(f"{'Missing values'.rjust(SPACE)} : {frame[column].isna().sum()}")
    
    # Numerical variables
    if frame[column].dtypes != 'O' and len(frame[column].unique()) > MAX_ORD:
        # Descriptive stats
        print(f"{'Correlation(target)'.rjust(SPACE)} : {frame[column].corr(frame[TARGET])}")
        print(f"{'Mean'.rjust(SPACE)} : {frame[column].mean()}")
        print(f"{'Min'.rjust(SPACE)} : {frame[column].min()}")
        print(f"{'Q1'.rjust(SPACE)} : {frame[column].quantile(0.25)}")
        print(f"{'Median'.rjust(SPACE)} : {frame[column].median()}")
        print(f"{'Q3'.rjust(SPACE)} : {frame[column].quantile(0.75)}")
        print(f"{'Max'.rjust(SPACE)} : {frame[column].max()}")
        print(f"{'Var'.rjust(SPACE)} : {frame[column].var()}")
        print(f"{'Std'.rjust(SPACE)} : {frame[column].std()}")
        
        stats, p_value = shapiro(frame[column].sample(n=MAX_SAMPLE))
        print(f"{'Normal distrib.'.rjust(SPACE)} : {round(stats, 3), round(p_value, 3)}")
        
        # Distribution
        plt.figure(figsize=(WIDTH, HEIGHT))
        sns.histplot(data=frame, x=column, hue=TARGET)
        plt.title(f"Distribution of {column}")
        plt.show()
        
        # Plot one against the other
        plt.figure(figsize=(WIDTH, HEIGHT))
        sns.boxenplot(data=train, x=TARGET, y=column)
        plt.title(f"{TARGET} vs {column}")
        plt.show()
    
    # Categorical variables
    else:
        # Count plot
        plt.figure(figsize=(WIDTH, HEIGHT))
        sns.countplot(data=train, x=column, hue=TARGET)
        plt.title(f"{TARGET} vs {column}")
        plt.show()
```
