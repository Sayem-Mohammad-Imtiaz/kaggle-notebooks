 The variable test preparation course is too subjective to be binary since what complete preparation means to some student can mean none to other so let's explore further using this variable to see if we can extract some useful insights

```python
df1 = df[df['test preparation course']=='none']
```

```python
lis = df['parental level of education'].unique()
average_len=[]
for i in lis:
    y=(len(df1[df1['parental level of education']==i ])/len(df[df['parental level of education']==i]))*100
    average_len.append(y)
plt.xlabel('percentage of students with test preparation none')
plt.ylabel('parental level of education')
sns.barplot(y=lis,x=average_len,palette=color_pallet)
plt.show()
average_len=[]
for i in lis:
    y=df1[df1['parental level of education']==i ]['average score'].mean()
    average_len.append(y)
plt.xlabel('average mean score with test preparation none')
plt.ylabel('parental level of education')
sns.barplot(y=lis,x=average_len,palette=color_pallet)
plt.show()
```

<div class="alert alert-block alert-info" style="font-size:15px; font-family:verdana; line-height: 2.0em;">
    üìåMost of the students whose parents went to high school didn't prepared for the test and scored lowest among all other groups<br>
    üìåEven though students whose parents had master degree mostly said they didn't prepared for the test even then they scored highest<br>
    üìåcompared to other groups students whose parents went to some high school mostly said that they have prepared for the test but even then scored less<br>
</div>

```python
lis = df['race/ethnicity'].unique()
average_len=[]
for i in lis:
    y=(len(df1[df1['race/ethnicity']==i ])/len(df[df['race/ethnicity']==i]))*100
    average_len.append(y)
plt.xlabel('percentage of students with test preparation none')
plt.ylabel('race/ethnicity')
sns.barplot(y=lis,x=average_len,palette=color_pallet)
plt.show()
average_len=[]
for i in lis:
    y=df1[df1['race/ethnicity']==i ]['average score'].mean()
    average_len.append(y)
plt.xlabel('average mean score with test preparation none')
plt.ylabel('race/ethnicity')
sns.barplot(y=lis,x=average_len,palette=color_pallet)
plt.show()
```

# Bonus

Lets train a simple machine learning model to predict average score

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
```

```python
df
```

```python
df_model = df.drop(['math score','reading score','writing score'],axis=1)
```

```python
y = df_model['average score']
df_model = df_model.drop('average score',axis=1)
```

```python
df_model = pd.get_dummies(df_model)
df_model
```

```python
x_train,y_train,x_test,y_test = train_test_split(df_model,y,test_size=0.2,random_state=42)
```

```python
model = Ridge()
model.fit(x_train,x_test)
pred = model.predict(y_train)
train_pred = model.predict(x_train)
score =  mean_squared_error(y_test,pred,squared=False)
score
```

<h1>Feature importance plot</h1>

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train,x_test)
feature_importance = np.array(model.feature_importances_)
feature_names = np.array(x_train.columns)
data={'feature_names':feature_names,'feature_importance':feature_importance}
df_plt = pd.DataFrame(data)
df_plt.sort_values(by=['feature_importance'], ascending=False,inplace=True)
plt.figure(figsize=(10,8))
sns.barplot(x=df_plt['feature_importance'], y=df_plt['feature_names'])
plt.xlabel('FEATURE IMPORTANCE')
plt.ylabel('FEATURE NAMES')
plt.show()
```
