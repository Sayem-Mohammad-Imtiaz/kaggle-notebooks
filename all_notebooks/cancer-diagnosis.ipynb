```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])

print(df['diagnosis'].head())
```

# TRAIN AND TEST SPLITTING

```python
x=df.drop(columns="diagnosis",axis=1)
y=df["diagnosis"]
```

## import required machine learning functions

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=0)
y_train.shape
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

# RANDOMFOREST CLASSIFIER

```python
from sklearn.ensemble import RandomForestClassifier
Model=RandomForestClassifier(max_depth=2)
Model.fit(x_train,y_train)
y_pred=Model.predict(x_test)

# Summary of the predictions made by the classifier
classfi=classification_report(y_test,y_pred)

acc=accuracy_score(y_pred,y_test)
acc
```

```python
conf=confusion_matrix(y_test,y_pred)
conf
```

# NAIVE BAYES

```python
from sklearn.naive_bayes import GaussianNB
Model = GaussianNB()
Model.fit(x_train,y_train)

y_pred=Model.predict(x_test)
acc=accuracy_score(y_pred,y_test)
acc
```

# KNeighbourClassifier

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=3)
model.fit(x_train,y_train)
y_pred = Model.predict(x_test)
acc=accuracy_score(y_pred,y_test)
acc
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(model,x_test,y_test)
```

# SUPPORT VECTOR MACHINES

```python
from sklearn.svm import SVC

Model = SVC()
Model.fit(x_train, y_train)

y_pred = Model.predict(x_test)
acc=accuracy_score(y_pred,y_test)
acc
```

# LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
model = LinearRegression()
model.fit(x_train,y_train)

y_pred= model.predict(x_test)
print('y_intercept      :',model.intercept_)
```

```python
print("number of coefficients    :",len(model.coef_))
```

```python
print("mean absolute error  :",metrics.mean_absolute_error(y_test,y_pred))
```

```python
print('Root Mean Sqrt Error RMSE   :',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
