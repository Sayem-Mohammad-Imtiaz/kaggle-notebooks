```python
op = pd.read_csv('/kaggle/input/instacart-market-basket-analysis/order_products__prior.csv')
orders = pd.read_csv('/kaggle/input/instacart-market-basket-analysis/orders.csv')
products = pd.read_csv('/kaggle/input/instacart-market-basket-analysis/products.csv')
```

1. Συμπληρώστε τα κενά στον παρακάτω κώδικα για να υπολογίσετε το μέγεθος κάθε παραγγελίας

```python
size = op.groupby('order_id')['product_id'].count().to_frame('order_size')
```

```python
size.head()
```

2. Συμπληρώστε τα κενά στον παρακάτω κώδικα για να υπολογίσετε τη πιθανότητα επαναγοράς ενός προϊόντος

```python
reorder = op.groupby('product_id')['reordered'].mean().to_frame('ratio')
```

```python
reorder.head()
```

3. Συμπληρώστε τα κενά στον παρακάτω κώδικα για να υπολογίσετε το πλήθος των παραγγελιών ανά μέγεθος καλαθιού

```python
size_results = size.groupby('order_size')['order_size'].count().to_frame('total_orders')
```

```python
size_results.head()
```

4. Συμπληρώστε τα κενά για να κρατήστε τα προϊόντα με περισσότερες από 30 αγορές.

```python
op30 = op.groupby('product_id').filter(lambda x: x.shape[0]>30)
```

```python
op30.head()
```

5. Συμπληρώστε τα κενά για να υπολογίστε την μέση σειρά αγοράς (τοποθέτησης στο καλάθι) κάθε προϊόντος για προϊόντα που έχουν αγοραστεί περισσότερες από 30 φορές.

```python
avg_pos = op30.groupby('product_id')['add_to_cart_order'].mean().to_frame('mean_add_to_cart_order')
```

```python
avg_pos.head()
```

6. Συμπληρώστε τα κενά για να ταξινομήσετε τα προϊόντα κατά φθήνουσα σειρά με βάση την μέση σειρά αγοράς τους σε όλες τις παραγγελίες. 

```python
avg_pos_asc = avg_pos.sort_values(by='mean_add_to_cart_order', ascending=True)
```

```python
avg_pos_asc.head(10)
```

7. Συμπληρώστε τα κενά για να βρείτε το όνομα του προϊόντος που αγοράζεται πιο γρήγορα κατά μέσο όρο

```python
a = avg_pos_asc.merge(products, on='product_id', how='left')
a.head()
```

9. Συμπληρώστε τα κενά για να δημιουργήστε ένα νέο πίνακα ενώνοντας τους πίνακες orders και op

```python
join = orders.merge(op, on='order_id', how='inner')
join.head()
```
