# About the dataset 
This dataset is derived and cleaned from the full OpenSky dataset to illustrate the development of air traffic during the COVID-19 pandemic. It spans all flights seen by the network's more than 2500 members 

1. callsign: the identifier of the flight displayed on ATC screens (usually the first three letters are reserved for an airline: AFR for Air France, DLH for Lufthansa, etc.)
2. number: the commercial number of the flight, when available (the matching with the callsign comes from public open API)
3. aircraft_uid: a unique anonymised identifier for aircraft;
4. typecode: the aircraft model type (when available);
5. origin: a four letter code for the origin airport of the flight (when available);
6. destination: a four letter code for the destination airport of the flight (when available);
7. firstseen: the UTC timestamp of the first message received by the OpenSky Network;
8. lastseen: the UTC timestamp of the last message received by the OpenSky Network;
9. day: the UTC day of the last message received by the OpenSky Network;
10. latitude_1, longitude_1, altitude_1: the first detected position of the aircraft;
11. latitude_2, longitude_2, altitude_2: the last detected position of the aircraft.
