### Twitter Climate Change Sentiment Dataset
* Tweets sobre mudança de clima coletados entre 27/04/2015 e 21/02/2018
* Dividido em 3 colunas: sentiment, message, tweetid
* sentiment pode ter 4 labels: 
1. 2 (news): o tweet é associado a notícias reais de mudança de clima
2. 1 (Pro):  o tweet apoia a crença de mudança de clima com causa nas ações do ser humano
3. 0 (Neutral): o tweet nem apoia nem refuta a ideia de mudança de clima com causa humana
4. -1 (Anti): o tweet desacrita na ideia de mudança de clima com causa humana
* message possui o texto do tweet

```python
print(climate.shape)
print()
print(climate.sample(10))
```
