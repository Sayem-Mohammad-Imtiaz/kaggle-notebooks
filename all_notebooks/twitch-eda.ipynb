```python
plt.xkcd()
plt.style.use('seaborn')
df['Channel'].unique()

import re

patt='[^a-zA-Z0-9\_]'#matches everything except alha,numbers and _ sign

#remove the name thet contains words other than english 
df['Channel']=df['Channel'].str.replace(patt,'',regex=True)
```

```python

px.bar(data_frame=df[:10],x='Channel',y='Watch time(Minutes)',title='Watch_times')
```

```python
#stream Times vs channel
px.bar(df.nlargest(10,columns='Stream time(minutes)'),x='Channel',y='Stream time(minutes)')
```

```python
val=df.groupby(['Language'])['Watch time(Minutes)'].max().sort_values(ascending=False).values
ind=df.groupby(['Language'])['Watch time(Minutes)'].max().sort_values(ascending=False).index

from plotly.subplots import make_subplots
fig=make_subplots(1,2,subplot_titles=('Max_Minutes_Watched Channel','Channel_Count'),specs=[[{"type": "xy"},{"type":"domain"}]])
fig1=px.bar(y=ind,x=val,color_discrete_sequence=['red'])
fig1.update_layout(xaxis_title='Minutes watched ',yaxis_title='Channel Languages')

language_cnt=df['Language'].value_counts().sort_values(ascending=False)
fig2=px.pie(names=language_cnt.index[:5],values=language_cnt.values[:5])
fig2.update_traces(textposition='inside', textinfo='value+label')

fig.add_trace(fig1.data[0],1,1)
fig.add_trace(fig2.data[0],1,2)
```

```python
#per million
df['Followers gained']=round(df['Followers gained']/1000000,2)
df['Views gained']=round(df['Views gained']/1000000,2)
df['Followers']=round(df['Followers']/1000000,2)
```

```python
px.scatter(df, x="Followers gained", y='Followers',size="Views gained", log_x=True, size_max=60,color='Channel')

#Both x and y axis are in million

#we can see that Fextralife ,eventhough has less followers gained highest views,whereas User Shroud with 7.74M followers gained only 30M Views
```

```python
fig=px.scatter(df['Stream time(minutes)'],y=df['Views gained'])
fig=px.scatter(df['Stream time(minutes)'],y=df['Followers gained'])
fig
#both views and follower gained doesnt depend on steam time
```

```python
#Top 5 youtubes based on watch times

fig=make_subplots(rows=2,cols=2,subplot_titles=('English top watched Channels','Koreaan top watched Channels','Russian top watched Channels','Spanish top watched Channels'),specs=[[{'type':'xy'},{'type':'xy'}],[{'type':'xy'},{'type':'xy'}]])
fig_eng=px.bar(df[df['Language']=='English'].sort_values(by='Watch time(Minutes)',ascending=False)[:10],x='Channel',y='Watch time(Minutes)',color_discrete_sequence=['lightgreen'])
fig_kor=px.bar(df[df['Language']=='Korean'].sort_values(by='Watch time(Minutes)',ascending=False)[:10],x='Channel',y='Watch time(Minutes)',color_discrete_sequence=['lightseagreen'])
fig_rus=px.bar(df[df['Language']=='Russian'].sort_values(by='Watch time(Minutes)',ascending=False)[:10],x='Channel',y='Watch time(Minutes)',color_discrete_sequence=['lightsalmon'])
fig_spa=px.bar(df[df['Language']=='Spanish'].sort_values(by='Watch time(Minutes)',ascending=False)[:10],x='Channel',y='Watch time(Minutes)',color_discrete_sequence=['lightpink'])

fig.add_traces(fig_eng.data[0],1,1)
fig.add_traces(fig_kor.data[0],1,2)
fig.add_traces(fig_rus.data[0],2,1)
fig.add_traces(fig_spa.data[0],2,2)
```
