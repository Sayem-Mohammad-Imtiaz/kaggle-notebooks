# Arranging the states on the basis of annual rainfall in descending order

# Creating a dataframe for Arunachal Pradesh

```python
dfap=df[['SUBDIVISION','YEAR','ANNUAL','Jan-Feb','Mar-May','June-September','Oct-Dec']]
ap=df.loc[ df['SUBDIVISION']=='Arunachal Pradesh']
```

```python
ap.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Observing the pattern of annual rainfall with time in Arunachal Pradesh

```python
sns.jointplot(x='ANNUAL',y='YEAR',data=ap,kind='hex')
```

```python
sns.distplot(x=ap["ANNUAL"])
```

```python
plt.figure(figsize=(15,9))
sns.heatmap(ap.corr())
```

```python
#analyzing rainfall pattern in Arunachal Pradesh during Winter months
sns.jointplot(x='YEAR',y='Jan-Feb',data=ap,kind='hex')
```

```python
#analyzing rainfall pattern in Arunachal Pradesh during Rainy months
sns.jointplot(x='YEAR',y='JUN',data=ap,kind='hex')
```

```python
#We observed that in Arunachal Pradesh more rainfall take place during rainy months rather than winter months.
#We also observed that the average rainfall in Arunachal Pradesh is decreasing with time , this may be becaue of deforestation and various other factors.
```

```python
plt.figure(figsize=(15,8))
sns.kdeplot(y='ANNUAL',x='YEAR',data=ap,shade=bool)
```

```python
#We observed that the annual rainfall in Arunachal Pradesh in gradually decreasing with time 
```

```python
plt.figure(figsize=(15,8))
sns.kdeplot(y='June-September',x='YEAR',data=ap,shade=bool)
```

```python
#Here we can clearly see that during rainiy months i.e June-September months, the annual rainfall is rapidly decreasing besides the fact that it receives most of its rainfall during rainy months
```

```python
#Analyzing the annual rainfall pattern in India
df.plot.hexbin(x='YEAR',y='ANNUAL',gridsize=23,figsize=(15,9),cmap=plt.get_cmap("Blues"))

plt.ylabel('ANNUAL RAINFALL')
plt.title("YEARLY RAINFALL IN INDIA")
```

```python
#We can clearly observe that the amount of rainfall remains same in India but the distribution changes like we saw in the case of Arunachal Pradesh and West Rajasthan.
```

```python
#Observing the most rainy month or duration in India
df.drop(['Latitude','Longitude','YEAR'],axis=1).plot(figsize=(15,6))
plt.legend(loc=(1,0),prop={'size': 12})
```
