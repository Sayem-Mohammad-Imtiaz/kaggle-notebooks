```python
import pandas as pd
import numba as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
data=pd.read_csv("../input/national-stock-exchange-dataset-sensex/nse_sensex.csv")
```

```python
data['DATE']=pd.to_datetime(data['DATE'])
```

```python
data['Day']=data['DATE'].dt.day_name()
```

```python
plt.figure(figsize=(20,10))
sns.barplot(data=data , x='SYMBOL',y='TRADES')
plt.xticks(rotation=90)
plt.title("Symbol with Trades")
plt.show()
```

```python
plt.style.use("fivethirtyeight")
fig,axis=plt.subplots(2,2 ,figsize=(20,10))
sns.lineplot(x='DATE',y='OPEN',data=data ,ax=axis[0,0])
sns.lineplot(x='DATE',y='CLOSE',data=data ,ax=axis[0,1])
sns.lineplot(x='DATE',y='HIGH',data=data ,ax=axis[1,0])
sns.lineplot(x='DATE',y='LOW',data=data ,ax=axis[1,1])
```

```python
data.groupby('SYMBOL').mean().sort_values(by='TRADES',ascending=False)
```

```python
data.groupby('Day').sum().sort_values(by='PREV CLOSE',ascending=False)
```

```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(20,10))
sns.lineplot(data=data , x='SYMBOL',y='OPEN')
sns.lineplot(data=data , x='SYMBOL',y='HIGH')
sns.lineplot(data=data , x='SYMBOL',y='LOW')
sns.lineplot(data=data , x='SYMBOL',y='CLOSE')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
plt.style.use("ggplot")
fig,axis=plt.subplots(2,2 ,figsize=(20,10))
sns.lineplot(x='DATE',y='TURNOVER',data=data ,ax=axis[0,0])
sns.lineplot(x='DATE',y='TRADES',data=data ,ax=axis[0,1])
sns.lineplot(x='DATE',y='DELIVERABLE VOLUME',data=data ,ax=axis[1,0])
sns.lineplot(x='DATE',y='%DELIVERBLE',data=data ,ax=axis[1,1])
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(data=data , x='DATE',y='VOLUME',color="yellow")
```
