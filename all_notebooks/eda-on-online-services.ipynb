> **Let's load Data and drop ID column as we already have index**

```python
df=pd.read_csv('/kaggle/input/50-online-services-dataset/Dataset of 50 Online Services Advertised in the Internet Marketing Forum searchengines.guru.csv')
df.drop('ID',axis=1,inplace=True)
df['User Forum Registration']=pd.to_datetime(df['User Forum Registration'])
df.head()
```

```python
df.columns
```

> ***We observed That There are 4 labels of Trustworthyness, So lets make them one by giving rating on Trustworthyness based on the Count of 'trustworthy' value in all labels.***

```python
df['TrustWorthyness Rating']=df[df=='trustworthy'].count(axis=1)
df.iloc[:5,22:29]
```

> ***With this lets find out Top 10 most Trust Worthy Forum:***

```python
df.nlargest(10,columns='TrustWorthyness Rating').loc[:,['User Forum Name','TrustWorthyness Rating']]
```

> **Now lets find the realtion between The Rating of the forum and the its Trustworthyness**

```python
plt.figure(figsize=(15,5))
plt.title('Trustworthyness Rating VS User Forum Rating',fontsize=20)
plt.xlabel('User Forum Rating',fontsize=12)
plt.ylabel('TrustWorthyness Rating',fontsize=12)
plt.scatter(df['User Forum Rating'],df['TrustWorthyness Rating'])
plt.show()
```

> **From the plot we can say that trustworthyness is not a very crutial factor for the Rating**

> **Now Lets Find the Distribution Rate for Various Columns So We Can More Understand These Columns**

```python
plt.figure(figsize=(20,10))
plt.subplot(2,3,1)
plt.title('Web Reputation: Cisco Umbrella Security Score Distribution')
plt.pie(df['Web Reputation: Cisco Umbrella Security Score'].value_counts().values,labels=df['Web Reputation: Cisco Umbrella Security Score'].value_counts().index)
plt.subplot(2,3,2)
plt.title('Web Reputation: BrightCloud® TI Risk Distribution')
plt.pie(df['Web Reputation: BrightCloud® TI Risk'].value_counts().values,labels=df['Web Reputation: BrightCloud® TI Risk'].value_counts().index)
plt.subplot(2,3,3)
plt.title('TrustWorthyness Rating Distribution')
plt.pie(df['TrustWorthyness Rating'].value_counts().values,labels=df['TrustWorthyness Rating'].value_counts().index)
plt.subplot(2,3,4)
plt.title('Type of User Advertising Distribution')
plt.pie(df['Type of User Advertising'].value_counts().values,labels=df['Type of User Advertising'].value_counts().index)
plt.subplot(2,3,5)
plt.title('searchengines.guru Sub Category Distribution')
plt.pie(df['searchengines.guru Sub Category'].value_counts().values,labels=df['searchengines.guru Sub Category'].value_counts().index)
plt.subplot(2,3,6)
plt.title('Service Overall Maturity Level Distribution')
plt.pie(df['Service Overall Maturity Level'].value_counts().values,labels=df['Service Overall Maturity Level'].value_counts().index)


plt.show()
```

*From the above visualizatoins we can conclude the following conclusions:*
1. Acc. to Cisco Umbrella Security Service's Web rating Maximum forums are meduim risky and very few are high risky which is quite good.
2. The maximum forum's Type of User Advertising is Organization based rather than Individual based.
3. The Maximum Forums are highly Mature and also rest are Medium Matures ,This is also nice.**

> **Let's Do Country Wise Analysis**

```python
hosting_country_df=df.groupby('Site Hosting Country')
plt.figure(figsize=(15,7))
plt.title('Countries with Number of Sites Hosted',fontsize=25)
plt.xlabel('Number of Sites Hosted',fontsize=15)
plt.ylabel('Counties',fontsize=15)
plt.barh(hosting_country_df.count()['User Forum Name'].sort_values().index,hosting_country_df.count()['User Forum Name'].sort_values().values,color='pink')
for index, value in enumerate(hosting_country_df.count()['User Forum Name'].sort_values().values):
    plt.text(value+0.1, index,str(value))
plt.show()
```

> **Lets Find Out On which Year Maximum Company Got Registered**

```python
df['Registration Year']=df['User Forum Registration'].dt.year
df['Registration Year'].head()
```

```python
year_df=df.sort_values(by='Registration Year').groupby('Registration Year')
plt.figure(figsize=(15,7))
plt.title('Number Of Forum Registered Yearly',fontsize=25)
plt.ylabel('Numbers of Forum registered',fontsize=15)
plt.xlabel('Years',fontsize=15)
plt.bar(year_df.count()['User Forum Name'].index,year_df.count()['User Forum Name'].values,color='yellow')
plt.xticks(year_df.count()['User Forum Name'].index)
for x in year_df:
    plt.text(x[0],x[1]['User Forum Name'].count()+0.1,str(x[1]['User Forum Name'].count()))
plt.show()
```
