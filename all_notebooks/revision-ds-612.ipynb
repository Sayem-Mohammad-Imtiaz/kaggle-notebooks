Extract outliers


```python
import matplotlib.pyplot as plt
desc = DataFrame.iterable.describe()
Q1 = desc[4]
Q2 = desc[5]
Q3 = desc[6]
print('Q1=', Q1, 'Q2=', Q2,'Q3=', Q3)
IQR = Q3-Q1
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
print("Anything outside this range is an outlier: (", lower_bound ,",", upper_bound,")")
DataFrame[DataFrame.iterable < lower_bound].iterable
print("Outliers: ",DataFrame[(DataFrame.iterable < lower_bound) | (DataFrame.iterable > upper_bound)].iterable.values)
boxplot=DataFrame.boxplot(column='iterable')
plt.show()
```

# Linear Regression

```python
#import library required for linear regression
from sklearn.linear_model import LinearRegression
#import library required for data visualization
import matplotlib.pyplot as plt
_,ax = plt.subplots(figsize=(5,5))
#select model
model = LinearRegression()
#select feature (predictor =age_n) 
x = DataFrame.age_n.values.reshape(-1,1)
#target ( response variable=income_n) 
y = DataFrame.income_n.values.reshape(-1,1)
model.fit(x,y)

ax.scatter(x, y, color = "red")
y_pred= model.predict(x).reshape(-1,1)

ax.scatter(x,y_pred, color = "blue")
plt.show()
```

```python
_, ax=plt.subplots(figsize=(15,5))

n_samples=50 #Y.shape[0]
n_samples=y.shape[0]
ax.scatter(x=range(0, n_samples),y=y[0:n_samples],c= 'blue',label='Actual')
ax.scatter(x=range(0, n_samples),y=y_pred.T[0:n_samples],c= 'red',label='predicted')

plt.title('Actual and predicted values')
plt.xlabel('Samples')
plt.ylabel('value')
plt.legend()
plt.show()
```
