### Modeling First DataFrame

```python
df_cleaned.head()
```

```python
X_1 = df_cleaned.drop('price',axis=1)
Y_1 = df_cleaned.price
x_train_1,x_test_1,y_train_1,y_test_1 = train_test_split(X_1,Y_1,random_state=1)
```

```python
model1 = LinearRegression()
model2 = RandomForestRegressor()
```

```python
model1.fit(x_train_1,y_train_1)
model2.fit(x_train_1,y_train_1)
```

```python
model1.score(x_test_1,y_test_1)
```

```python
model2.score(x_test_1,y_test_1)
```

Absolutely Random Forest Regressor Algorithm is performing Best
