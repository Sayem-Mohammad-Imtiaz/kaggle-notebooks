*I will apply Kmeans and Hierarchial Clustering with this dataset. These two methods look like each other. Both of them has no information of data labels. One difference is when Kmeans decide optimum cluster, it consider wcss(shows optimum k) but Hierarchial clustering consider ward(to minimum variant clustering).Firstly we define dataset and variable.*

```python
data=pd.read_csv("/kaggle/input/heart-disease-uci/heart.csv")
data.head()

y=data["age"].values
x1=data["trestbps"].values
x2=data["chol"].values

x=np.concatenate((x1,x2),axis=0)
```

After defining variable, it is gonna be good to see plots.

```python
#visualization
plt.scatter(x1,y)
plt.scatter(x2,y)
plt.show()
```

1. *As we see, there are two groups. To detect optimum k,we will predict with Kmeans Clustering.*

```python
#KMeans Clustering
from sklearn.cluster import KMeans
wcss=[] #wcss, provide finding optimum 'k' clustering number.
for k in range(1,15):
    Kmeans=KMeans(n_clusters=k)
    Kmeans.fit(data)
    wcss.append(Kmeans.inertia_) #inertia: For each k value, shows wscc values. 
plt.plot(range(1,15),wcss)
plt.xlabel("number of k values")
plt.ylabel("wcss")
plt.show()
```

On the graphic we can see elbow. So elbow point will define our cluster number. There is no sharply elbow sign but i think we can choose '3'.

```python
#model for k=2

kmeans2=KMeans(n_clusters=2)
clusters=kmeans2.fit_predict(data)

data["label"]=clusters #we define labels for clusters and we will be completed classification.
plt.scatter(data["chol"],data["trestbps"],c = clusters)
plt.xlabel("trestbps")
plt.ylabel("chol")
plt.show()    

```
