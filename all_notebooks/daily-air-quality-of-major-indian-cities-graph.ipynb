```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (32, 12)
plt.rcParams["font.size"] = 24

f = open("/kaggle/input/daily-air-quality-dataset-india/air_quality_index_india.csv", "r")

tarr = ['Hyderabad', 'Bhopal', 'Chandigarh', 'Delhi', 'New Delhi', 'Nashik', 'Chennai', 'Thiruvananthapuram', 'Bengaluru', 'Jaipur', 'Mumbai', 'Muzaffarnagar', 'Patna', 'Kolkata', 'Visakhapatnam', 'Gh훮zi훮b훮d', 'Lucknow', 'Gandhinagar', 'H훮pur', 'Shillong', 'Thrissur', 'Mysore', 'Nagpur']

outs = [0]*len(tarr)

lines = f.readlines()

for i in range(1, len(lines)):
    line = lines[i].replace("\n", "").split(",")
    
    outs[tarr.index(line[2])] += float(line[3])

for i in range(0, len(tarr)):
    tarr[i] = [tarr[i], outs[i]]

tarr.sort(key=lambda xx: xx[1], reverse=True)

x = []
y = []

for i in range(0, len(tarr)):
    x.append(tarr[i][0])
    y.append(tarr[i][1])

plt.title("Daily Air Quality of Major Indian Cities")
plt.xlabel("City")
plt.ylabel("The Average daily recorded value of pm25. Amount by city.")
plt.grid(linestyle="--")
plt.xticks(rotation=75)
plt.plot(x, y, label="Value", linewidth=2.61, marker="o")
plt.legend()
plt.show()
```
