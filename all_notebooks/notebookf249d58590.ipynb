```python
import pandas as pd
import numba as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
data=pd.read_csv("../input/apple-stock-prices-20152020/AAPL.csv")
```

```python
data['date']=pd.to_datetime(data['date'])
```

```python
data["Day"]=data['date'].dt.day_name()
```

```python
plt.style.use("dark_background")
fig,axis=plt.subplots(2,2, figsize=(20,10))
sns.lineplot(x='date',y='close',data=data,color='red',ax=axis[0,0])
sns.lineplot(x='date',y='high',data=data,color='red',ax=axis[0,1])
sns.lineplot(x='date',y='low',data=data,color='red',ax=axis[1,0])
sns.lineplot(x='date',y='open',data=data,color='red',ax=axis[1,1])
```

```python
data.set_index('date',inplace=True)
```

```python
df2015=data.loc['2015']
df2016=data.loc['2016']
df2017=data.loc['2017']
df2018=data.loc['2018']
df2019=data.loc['2019']
df2020=data.loc['2020']
```

```python
plt.style.use("dark_background")
fig,axis=plt.subplots(3,2, figsize=(30,15))
sns.lineplot(x='date',y='volume',data=df2015,color='red',ax=axis[0,0])
sns.lineplot(x='date',y='volume',data=df2016,color='red',ax=axis[0,1])
sns.lineplot(x='date',y='volume',data=df2017,color='red',ax=axis[1,0])
sns.lineplot(x='date',y='volume',data=df2018,color='red',ax=axis[1,1])
sns.lineplot(x='date',y='volume',data=df2019,color='red',ax=axis[2,0])
sns.lineplot(x='date',y='volume',data=df2020,color='red',ax=axis[2,1])
```

```python
plt.style.use("dark_background")
fig,axis=plt.subplots(2,2, figsize=(20,10))
sns.lineplot(x='date',y='adjClose',data=data,color='white',ax=axis[0,0])
sns.lineplot(x='date',y='adjHigh',data=data,color='white',ax=axis[0,1])
sns.lineplot(x='date',y='adjLow',data=data,color='white',ax=axis[1,0])
sns.lineplot(x='date',y='adjOpen',data=data,color='white',ax=axis[1,1])
```

```python
plt.style.use("dark_background")
plt.figure(figsize=(20,10))
sns.lineplot(x='date',y='adjVolume',data=data,color='white')
```
