```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../input/boston-housing/boston_housing.csv')
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Check the spread of data across classes
sns.kdeplot(x='medv', data=df)
```

```python
sns.heatmap(df.corr(), cmap='Accent')
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X = df[df.columns[:-1]] # Everything except medv (the outcome)
y = df['medv']

scaler = preprocessing.StandardScaler().fit(X)
X = scaler.transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1) # total: 768 values
print('Train :', len(X_train))
print('Test  :', len(X_test))
```

```python
log = LinearRegression()
log.fit(X_train, y_train)

log.score(X_test, y_test)
```
