```python
dataset = pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(13,7))
sns.swarmplot(x=dataset['test preparation course'], y=dataset['math score'], hue=dataset['gender'], size=7, palette='RdYlBu')
plt.title('Math Score by Preparation Course', pad=20, fontsize=16)
plt.xlabel('Test Preparation Course', fontsize=14)
plt.ylabel('Math Score', fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(13,7))
sns.swarmplot(x=dataset['test preparation course'], y=dataset['writing score'], hue=dataset['gender'], size=7, palette='RdYlBu')
plt.title('Writing Score by Preparation Course', pad=20, fontsize=16)
plt.xlabel('Test Preparation Course', fontsize=14)
plt.ylabel('Writing Score', fontsize=14)
plt.show()
```
