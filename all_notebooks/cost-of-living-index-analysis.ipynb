```python
def battle_of_two(x,y,title,plot='scatter'):
    plt.figure(figsize=(7,7))
    subset = df[[x, y]]
    plt.figure(figsize=(7,7))
    if plot=='scatter':
        scatterplot(data=subset,x=x,y=y)
        plt.title(title)
        plt.show()
    else:
        lineplot(data=subset,x=x,y=y)
        plt.title(title)
        plt.show()
    
    #Correlation Plot
    plt.figure(figsize=(7,7))
    heatmap(subset.corr(),fmt='g',annot=True)
    plt.title('Correlation HeatMap')
    plt.show()
```

# **Restaurant Price Index vs  	Local Purchasing Power Index**

```python
battle_of_two('Restaurant Price Index','Local Purchasing Power Index','Restaurant Price Index vs Local Purchasing Power Index')
```
