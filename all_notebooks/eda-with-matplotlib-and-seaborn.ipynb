# Analyzing different types of search by users with diffferent variables

```python
plt.figure(figsize=(15,7))
sns.barplot(x='Administrative',y='Revenue',data=df)
plt.xlabel('Administrative(no. of pages user visited)')
```

```python
#Hence we can clearly see that revenue increases with increase of number of administrative pages visited
```

```python
plt.figure(figsize=(15,7))
sns.barplot(y='Administrative_Duration',x='Revenue',data=df,hue='Month')
plt.legend(loc=(1,0),prop={'size':25.4})
```

```python
#In the month of February the duration of administrative searches are huge  
```

```python
df.head(2)
```

```python
plt.figure(figsize=(15,7))
sns.barplot(x='Informational',y='Revenue',data=df)
plt.xlabel('Informational(no. of pages user visited)')
```

```python
#here we can see that number of informational pages visited is not that huge,may be due to the fact that the required information has been received in the previous pages
#Also we can observe that after certain number of pages the revenue gradually increases and there are few outliers
```

```python
plt.figure(figsize=(15,7))
sns.barplot(y='Informational_Duration',x='Revenue',data=df,hue='Month')
plt.legend(loc=(1,0),prop={'size':25.4})
```

```python
#Revenue is high for informational searches especially in the month of March and July 
```
