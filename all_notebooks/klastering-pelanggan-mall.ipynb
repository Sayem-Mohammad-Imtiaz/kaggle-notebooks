```python

### Pemilihan fitur untuk model
#Mempertimbangkan hanya 2 fitur (Pendapatan tahunan dan Skor Pengeluaran)
X= dataset.iloc[:, [3,4]].values
```

```python
from sklearn.cluster import KMeans
wcss=[]
```

```python
for i in range(1,11):
    kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
kmeansmodel = KMeans(n_clusters= 5, init='k-means++', random_state=0)
y_kmeans= kmeansmodel.fit_predict(X)
```

```python
#Visualisasi semua kluster
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'purple', label = 'Klaster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'pink', label = 'Klaster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'red', label = 'Klater 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'green', label = 'Klaster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'blue', label = 'Klaster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Sentroid')
plt.title('Klaster Pelanggan')
plt.xlabel('Pendapatan tahunan (k$)')
plt.ylabel('Skor Pembelanjaan (1-100)')
plt.legend()
plt.show()
```

```python

### Interpretasi Model
#Klaster 1 (Ungu)   -> Penghasilan tinggi tetapi Pembelanjaan lebih sedikit
#Klaster 2 (pink)   -> Rata dalam hal Penghasilan dan Pembelanjaan
#Klaster 3 (Merah)  -> Penghasilan tinggi dan juga Pembelanjaan banyak 
#Klaster 4 (Hijau)  -> Penghasilan lebih sedikit tetapi Pembelanjaan lebih banyak
#Klaster 5 (Biru)   -> Penghasilan lebih sedikit, Pembelanjaan lebih sedikit
```
