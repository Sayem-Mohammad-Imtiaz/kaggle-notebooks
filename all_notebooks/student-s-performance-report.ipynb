1. Performance of students(group by students) in all exams (Sheet 1 : Relation between scores and gender)

    * In reading - females with an average of **72.6%** and males with an average of **65.4%** 
    * In writing - females with an average of **72.4%** and males with an average of **63.3%**

2. What patental level of education does parents possess?(Sheet 2 : Patental level of education)

    * `Master's degree` is the least comman amomg parents.
    
3. Which group(race) has highest and lowest students?(Sheet 3 : Comparision among various groups)
    * Highest - `group C`
    * Lowest - `group A`
    
3. What is the status of students for test preparation?(Sheet 4 : Gender vs test preparation)
    * 64.2% students have `none` in their preparation status having the total of **females = 334** and **males = 308**.
    * 35.8% students have `completed` in their preparation status having the total of **females = 184** and **males = 174**.
   
4. Relationship between lunch and marks scored by students.(Dashboard : Relation between lunch and marks scored by students)
    * 64.5% lunch given to students is `standard` and rest 35.5% lunch is `free` to students. 
    * Students getting `standard` lunch are doing better than students getting `free` lunch w.r.t all subjects.
