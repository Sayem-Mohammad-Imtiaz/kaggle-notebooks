```python
import pandas as pd 
import numpy as np 
import plotly as pty 
import plotly.express as pex
import matplotlib.pyplot as plt 


street_crimes = pd.read_csv('../input/crimes-in-boston/crime.csv' ,encoding='windows-1252' )

street_crimes = street_crimes[["STREET" , "INCIDENT_NUMBER", "DISTRICT" ]]

# group the district and streets together. 
#scm = street_crimes.groupby(["DISTRICT" , "STREET" ]).count()

# reset the index and get the values out. 
#scm = scm[['INCIDENT_NUMBER']] ).reset_index().values

# build this into a dataframe and add colums to this data frame. . 
#scm = pd.DataFrame(data = (scm) , columns=["DISTRICT", "", "STREET"))


# together 
#scm = pd.DataFrame(data = (street_crimes.groupby(["DISTRICT" , "STREET" ]).count()[['INCIDENT_NUMBER']] ).reset_index().values , columns=["DISTRICT", "", "STREET" ]   ) 


scm = pd.DataFrame(
    data = (
        street_crimes.groupby(
            [
                "DISTRICT" , "STREET" 
            ]
        ).count()[
            [
                'INCIDENT_NUMBER'
                ]
            ] 
        ).reset_index().values , columns=[
            "DISTRICT", "",  "STREET" 
        ]   
    ) 

#scm = scm.groupby(["DISTRICT"]).nlargest(7)


#labels = ["DISTRICT" , "STREET"]
fig = pex.sunburst(scm , path = ["DISTRICT",  "STREET" ], values = "STREET" )

fig.show()

plt.show()
```
