Import the pandas, meaplotlib, and seaborn libries to allow them to be used throughout the code.

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import numpy as np
```

To start off we have to initialise the dataset. This is done by passing the dataset url into the pd.read_csv function which converts the data into something easy to read by pandas. To use this data in the future it is stored in a variable named data.

To show the first 5 items in the dataset data.head function is called. data.tail does the same thing but shows the last 5 items in the dataset.

```python
path = '../input/roadcrashdata2019/2019_DATA_SA_Casualty.csv'
data = pd.read_csv(path, index_col="REPORT_ID", parse_dates=True)

data.head()
```

The data below demonstrates the percentage of types of poeple that were involved in casualties.

From the 'Casualty Type' column I get the counts for each of the types. I them plot these onto a pie chart using .plot.pie(). The paramiter given is one i found online shows the percentage for each area.

The chart demonstrates that 58% of poeple involved in crashes are drivers. The other groups are passengers, the next biggest portion. This would be because many drivers have pasangers in the car when they crash. Riders account for 16% which is much more than the 4.5% of vehicles on the road they are. This shows that motorbikes are much more prone to crashing than cars are. pedestrians are also not very likely to be hit by a car by their 5.3% portion.

```python
casdata = data["Casualty Type"].value_counts()
casdata.plot.pie(autopct="%.1f%%")
# sns.barplot(x=casdata.index, y=casdata)
plt.title("Percentages of Casualty types")
plt.ylabel("")
```

This data below shows the relationship between the age of poeple involved in casualties and how many popele were involved in these casualties. 

After sorting the data by the 'AGE' column I get the data from the 'AGE' column and count how many of each of the values there are. Storing this as the variable 'agedata' get the index (count) on the y axis and the ages on the x axis. I dont use the last item in these however because they demonstrate the unknown ages, which dont help in this instance.

This graph shows that the least likely poeple to be involved in a crash are those younger than 15 and older than 75. At 16, going into the 20's the casualties are at their maximum with over 130 casualties occuring per age. The reason that this is the max would be beucase these people are the least experienced at driving. They are also the most suseptable to  drig/drink driving because they are irresponsable. After this the crashes per age steadily decrease.

```python
data = data.sort_values(["AGE"])
agedata = data["AGE"].value_counts().sort_index(ascending=True)
plt.figure(figsize=(20, 12))
plt.xticks(rotation=80)
plt.title("Age vs crash counts")
plt.ylabel('Crash count')
# sns.barplot(y=agedata[:-1], x=agedata.index[:-1])
sns.lineplot(y=agedata[:-1], x=agedata.index[:-1])
```
