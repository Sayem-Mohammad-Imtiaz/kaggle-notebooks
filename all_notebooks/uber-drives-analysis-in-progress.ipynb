# **EDA(Null Values and Dropping Columns)**

**Checking for Null Values in data**

```python
sns.heatmap(train.isnull(),cmap = 'viridis')
plt.show()
print(train.isnull().sum())
print('')
print('Percentage of Null Value in Purpose column :',train['PURPOSE*'].isnull().sum()/len(train['PURPOSE*'])*100,'%')
```

Since 43% of values is null we can drop that column

**Droping Purpose Column**

```python
train.drop('PURPOSE*',axis = 1, inplace = True)
```

**Dropping rest of the null Values**

```python
train.dropna(inplace=True)
```

# **EDA : Start And Stop Time Analysis**

**Convering Start and Stop Time to Date Time Object**

```python
train['Start'] = pd.to_datetime(train['START_DATE*'])
train['End'] = pd.to_datetime(train['END_DATE*'])
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}

train['s_month'] = train['Start'].apply(lambda x : x.month)
train['s_day'] = train['Start'].apply(lambda x : x.dayofweek)


train['s_day'] = train['s_day'].map(dmap)
```

```python
train.drop('START_DATE*',axis = 1,inplace = True)
train.drop('END_DATE*',axis = 1,inplace = True)
```

```python
train.head(10)
```

**Creating a new Column for Time Travelled**

```python
train['Time_travelled'] = train['End'] - train['Start'] 
train['Time_travelled'] = train['Time_travelled'].apply(lambda x: x.days/(24*60) +x.seconds/60)
```
