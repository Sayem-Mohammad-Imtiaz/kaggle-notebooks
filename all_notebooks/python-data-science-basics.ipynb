```python
#what if there is no local scope
x=2
def f():
    y=2*x       #no local scope x is defined in f
    return y
print(f())      #How will f calculate y? It will use the global scope x, since there is no local.
```

```python
#nested functions
def square():
    """returns square of a precalculated value"""
    def add():
        """returns sum of two defined local values"""
        x=3
        y=2
        z=x+y
        return z
    return add()**2
print(square())
```

**RESAMPLING PANDAS TIME SERIES**

Resampling: statistical method over different time intervals

Needs string to specify frequency like "M" = month or "A" = year

Downsampling: reduce date time rows to slower frequency like from daily to weekly

Upsampling: increase date time rows to faster frequency like from daily to hourly

Interpolate: Interpolate values according to different methods like ‘linear’, ‘time’ or index’

https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.interpolate.html

```python
data1.resample("A").mean()
```

```python
data1.resample("M").mean()
```

**PIVOTING DATAFRAMES**

Pivoting is reshaping.

```python
dict1={"treatment":["A","A","B","B"],"gender":["F","M","F","M"],"response":[10,45,5,9],"age":[15,4,72,65]}
df=pd.DataFrame(dict1)
df
```

```python
df1=df.pivot(index="treatment",columns="gender",values="response")
df1
```

**STACKING AND UNSTACKING**

**Unstack:** Pivot a level of the (necessarily hierarchical) index labels.

Returns a DataFrame having a new level of column labels whose inner-most level consists of the pivoted index labels.

```python
df1=df.set_index(["treatment","gender"])
df1
```

```python
#let's unstack it
df1=df1.unstack(level=0)
df1
```

```python
df1.index
```

```python
df1.columns
```

```python
df1.unstack(level=1)
```

```python
df2=df1.swaplevel(0,1)
df2
```

**MELTING**

Unpivot a DataFrame from wide to long format, optionally leaving identifiers set.

This function is useful to massage a DataFrame into a format where one or more columns are identifier variables (id_vars), while all other columns, considered measured variables (value_vars), are “unpivoted” to the row axis, leaving just two non-identifier columns, ‘variable’ and ‘value’.

```python
melted=pd.melt(df,id_vars="treatment",value_vars=["response","age"])
melted
```
