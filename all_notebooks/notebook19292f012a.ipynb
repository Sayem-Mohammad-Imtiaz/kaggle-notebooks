```python
import pandas as pd 
import numpy as np
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
N = pd.read_csv('../input/2020-united-states-presidential-election/candidates_info.csv')
N
N = N \
    .rename(columns = 
           {'city of residence':'city_of_residence',
            'state of residence':'state_of_residence',
              
           })
N
```

```python
group= N.groupby('city_of_residence').agg({'party': 'count'})

group
```

```python
sex= N.groupby('sex').agg({'party': 'count'})
sex
```

```python
party= N.groupby('party').agg({'sex': 'count'})
party
```

```python
children= N.groupby('children').agg({'sex': 'count'})
children
```

```python
state_of_residence= N.groupby('state_of_residence').agg({'sex': 'count'})

state_of_residence
```

```python
p= N.groupby('party').agg({'children': 'sum'})
p
```
