### Questão 1 - Item C - Representação Gráfica
Para cada uma das variáveis, produza um ou mais gráficos, usando matplotlib, que descreva seu comportamento / caracteristica. Lembre-se que estes gráficos precisam ser compatíveis com a classificação da variável. 

### 1. Quantidade de Municípios por UF

```python
MunUF = df[['uf','nome_municipio']]

MunporUF = MunUF.groupby(by='uf',as_index=False).count().sort_values(by=['nome_municipio'], ascending=False)
MunporUF
x = MunporUF['uf']
y = MunporUF['nome_municipio']
plt.figure(figsize=(20, 10))
plt.bar(x, y)
plt.title('Quantidade de Municípios por Estado')
plt.xlabel('Estados')
plt.ylabel('Nº Municípios')

plt.show()
```

### 2. Total de Eleitores por UF

```python
EleitoresUF = df[['uf','total_eleitores']]

EleitporUF = EleitoresUF.groupby(by='uf',as_index=False).sum().sort_values(by=['total_eleitores'], ascending=True)
EleitporUF

x = EleitporUF['uf']
#Dividindo o número de eleitores por milhão para que o gráfico fique melhor visualizado
y = EleitporUF['total_eleitores']/1000000
plt.figure(figsize=(20, 10))
plt.barh(x, y)
plt.title('Total de Eleitores por Estado por milhão de Eleitores')
plt.ylabel('Estados')
plt.xlabel('Nº Eleitores (em milhões)')
plt.show()
```

### 3. Distribuição de Eleitores por gênero

```python
#Dataframe com apenas as informações relevantes para o gráfico
GenUF = df[['uf','gen_feminino','gen_masculino','gen_nao_informado']]

#Realizando a soma por gênero
soma_feminino = GenUF['gen_feminino'].sum(axis = 0)
soma_masculino = GenUF['gen_masculino'].sum(axis = 0)
soma_naoinformado = GenUF['gen_nao_informado'].sum(axis = 0)
Total = soma_feminino + soma_masculino + soma_naoinformado
perc_feminino = round((soma_feminino/Total)*100,2)
perc_masculino = round((soma_masculino/Total)*100,2)
perc_naoinformado = round((soma_naoinformado/Total)*100,2)

#Construindo o gráfico de Pizza
nomes = 'Feminino','Masculino','Sexo Não Informado'
tamanho = [perc_feminino,perc_masculino,perc_naoinformado]
explode = (0.1,0.1,0.1)

fig1, ax1 = plt.subplots()
ax1.pie(tamanho, explode=explode, labels=nomes, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
plt.title('Distribuição de Eleitores por Gênero')
plt.show()
```

### 4. Distribuição de Eleitores por gênero/UF

```python
GenUF.head()
GenporUF = GenUF.groupby(by='uf',as_index=False).sum().sort_values(by=['uf'], ascending=False)
GenporUF['gen_feminino'] = round(GenporUF['gen_feminino']/1000)
GenporUF['gen_masculino'] = round(GenporUF['gen_masculino']/1000)
GenporUF['gen_nao_informado'] = round(GenporUF['gen_nao_informado']/1000)

#Valores dos grupos e categorias
UF = GenporUF['uf']
Feminino = GenporUF['gen_feminino']
Masculino = GenporUF['gen_masculino']
NaoInformado = GenporUF['gen_nao_informado']
    
#Adicionando cores
colors = ['#00BFFF','#C1FFC1','#FFDEAD']
# Posição das barras no eixo X
r = range(len(UF))
barWidth = 1
#plot bars
plt.figure(figsize=(20,20))
ax1 = plt.bar(r, Feminino, color=colors[0], edgecolor='white', width=barWidth, label="Feminino")
ax2 = plt.bar(r, Masculino, bottom=np.array(Feminino), color=colors[1], edgecolor='white', width=barWidth, label='Masculino')
ax3 = plt.bar(r, NaoInformado, bottom=np.array(Feminino)+np.array(Masculino), color=colors[2], edgecolor='white', width=barWidth, label='Gênero não informado')
plt.legend(fontsize = 15)
# Customizando o eixo X
plt.xticks(r, UF, fontweight='bold')
plt.ylabel("Número de Eleitores (em milhares)",fontsize = 15)
plt.xlabel("Estados",fontsize = 15)

#Adicionando os valores para cada barra
for r1, r2, r3 in zip(ax1, ax2, ax3):
    h1 = r1.get_height()
    h2 = r2.get_height()
    h3 = r3.get_height()
    plt.text(r1.get_x() + r1.get_width() / 2., h1 / 2., "%d" % h1, ha="center", va="center", color="black", fontsize=10, fontweight="bold")
    plt.text(r2.get_x() + r2.get_width() / 2., h1 + h2 / 2., "%d" % h2, ha="center", va="center", color="black", fontsize=10, fontweight="bold")
    plt.text(r3.get_x() + r3.get_width() / 2., h1 + h2 + h3 / 2., "%d" % h3, ha="center", va="center", color="black", fontsize=10, fontweight="bold")
plt.savefig("stacked1.png")
plt.show()
```
