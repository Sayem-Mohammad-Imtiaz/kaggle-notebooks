```python
import pandas as pd
vix_data = pd.read_csv('/kaggle/input/indian-stock-index-eod-data1990-onwards/data_1990_2020/index_data/INDIA_VIX.csv');
vix_data.tail()
```

```python
vix_data['Date'] = pd.to_datetime(vix_data['Date'])
vix_data = vix_data[vix_data.Date > "2010-12-01"]
vix_data.head()
```

```python
vix_data.loc[vix_data['High']==0, ['Open','High','Low']] = vix_data[vix_data['High'] == 0].Close
vix_data
```

```python
def get_diffTimeFrame(data, timeframe):
    ohlc_dict = {'Open':'first', 'High':'max', 'Low':'min', 'Close': 'last'}
    result = data.resample(timeframe, on='Date').apply(ohlc_dict)
    result['Avg'] = (result['High'] + result['Low'] + result['Open'] + result['Close'])/4;
    return result
```

```python
def get_diffTimeFramePos(data, timeframe):
    ohlc_dict = {'Open':'first', 'High':'max', 'Low':'min', 'Close': 'last'}
    result = data.resample(timeframe, on='Date').apply(ohlc_dict)
    result['Range%'] = (result['High'] - result['Open']);
    return result
```

```python
def get_diffTimeFrameNeg(data, timeframe):
    ohlc_dict = {'Open':'first', 'High':'max', 'Low':'min', 'Close': 'last'}
    result = data.resample(timeframe, on='Date').apply(ohlc_dict)
    result['Range%'] = (result['Low'] - result['Open']);
    return result
```

```python
vix_weekly = get_diffTimeFrame(vix_data, 'W');
vix_weekly.tail(12)
```

```python
vix_weekly_pos = get_diffTimeFramePos(vix_data, 'W');
vix_weekly_neg = get_diffTimeFrameNeg(vix_data, 'W');
#vix weekly signed range distribution
vix_weekly_signed = vix_weekly_pos['Range%']
vix_weekly_signed = vix_weekly_signed.append(vix_weekly_neg['Range%'], ignore_index=True);
vix_weekly_signed.hist(figsize=[12,8],bins=95)
```

```python
#vix weekly range distribution
vix_weekly['High'].hist(figsize=[12,8],bins=85)
```

```python
#vix weekly range distribution
vix_weekly['Avg'].hist(figsize=[12,8],bins=85)
```

```python
vix_monthly = get_diffTimeFrame(vix_data, 'M');
vix_monthly.tail(12)
```

```python
vix_monthly_pos = get_diffTimeFramePos(vix_data, 'M');
vix_monthly_neg = get_diffTimeFrameNeg(vix_data, 'M');
#vix monthly signed range distribution
vix_monthly_signed = vix_monthly_pos['Range%']
vix_monthly_signed = vix_monthly_signed.append(vix_monthly_neg['Range%'], ignore_index=True);
vix_monthly_signed.hist(figsize=[12,8],bins=95)
```

```python
#vix monthly range distribution
vix_monthly['High'].hist(figsize=[12,8],bins=45)
```

```python
#vix monthly range distribution
vix_monthly['Avg'].hist(figsize=[12,8],bins=45)
```
