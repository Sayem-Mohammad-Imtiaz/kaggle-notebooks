```python
col=['Postal Code']
data_new = data.drop(columns=col,axis=1)
```

```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category', data=data_new)
plt.show()
```

```python
data_new['State'].value_counts()
```

```python
plt.figure(figsize=(15,15))
sns.countplot(x=data_new['State'])
plt.xticks(rotation=90)
plt.show()
```

```python
data['Sub-Category'].value_counts()
plt.figure(figsize=(15,15))
sns.countplot(x=data_new['Sub-Category'])
plt.xticks(rotation=180)
plt.show()
```

```python
sns.heatmap(data.corr(),annot=True)
plt.show()
```

```python
sns.heatmap(data.cov(),annot=True)
plt.show()
```

```python
plt.figure(figsize=(20,20))
sns.barplot(x=data_new['Sub-Category'],y=data_new['Profit'])
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot('Discount','Profit', data=data_new , color='y',label='Discount')
plt.legend()
plt.show()
```

```python
figsize=(15,10)
sns.pairplot(data_new,hue='Sub-Category')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10 , 6))
ax.scatter(data_new["Sales"] , data_new["Profit"])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```
