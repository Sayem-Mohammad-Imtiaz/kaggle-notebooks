#### Missing Valuse Relations

```python
# we will try to find the percentage of missing values in the each feature
dataset= data.copy()
# we will get the all the feature with missing values
feature_nan = [feature for feature in dataset.columns if data[feature].isnull().sum()>1]

# missing value percentage the ecah featrue
for feature in feature_nan:
    print("The missing value percentage(%) in ",feature,data[feature].isnull().mean())
```

 We can obsereve that in some features very less numer of missing value (less than 1 %) is present

```python
#plot the missing values on the heatmap
plt.figure(figsize=(12,6))
sns.heatmap(data.isnull(),cbar = False)
```

This gives us more detail about the features with the missing values. Features like GDP,Population,Hepatitis B,Alcohol,total expenditure have high number of missing values.

```python
# We can also see the count of missing value by count plot
for feature in feature_nan:
    dataset = data.copy()
    # Let's make a variable that indicates 1 if nan value present otherwise 0
    dataset[feature] = np.where(data[feature].isnull(),1,0)
    sns.countplot(x=feature,data = dataset,)
    plt.show()
```

```python
# There are very less number of missing value in the tagetr feature (Life expectancy) so we can drop these records
print(data.shape)
data = data.dropna(subset = ['Life expectancy '])
data.reset_index(drop =True,inplace= True)
print(data.shape)
```

### Numerical Features Relation with Life Expentency
1. Continuous variable
2. Discrete Variable

```python
num_features = [feature for feature in data.columns if data[feature].dtype != 'O']
data[num_features].head()                    
```

```python
# Lets find the discrete features from the data
discrete_features = [feature for featue in data.columns if len(data[feature].unique()) < 25]
discrete_features
```

we don't have any discrete numerical feature

```python
# lets see the relation between life expectancy and continuous numericalfeatures
for feature in num_features:
    dataset = data.copy()
    sns.jointplot(x='Life expectancy ',y = feature,data = dataset )
    plt.show()
```

```python
# Avgerage life ecpectancy every year
dataset.groupby(['Year'])['Life expectancy '].mean().sort_values()
```

Average Life expectancy is increasing with the years

```python
# How does Infant and Adult mortality rates affect life expectancy?

sns.scatterplot(x = 'Adult Mortality',y ='Life expectancy ',data = data )
```

We can that as the Adult mortality increases ,life expectancy decreases. So there should be a negative correlation between them.

```python
# How does Alcohol affect life expectancy? (Alcohol, recorded per capita (15+) consumption (in litres of pure alcohol))

sns.scatterplot(x = 'Alcohol',y ='Life expectancy ',data = data)
```

There is some positive correlation between alcohol consumption and life expectancy.
We can observe the there are some outlies present.Because for few records those having high alcohol consumption have life expectancy more than 65 and  for few records those having low alcohol consumption(less than 5) have life expectancy less than 45.  

```python
# What is the impact of schooling on the lifespan of humans?
sns.regplot(x ='Schooling',y ='Life expectancy ',data = data,marker = '+',color = 'g')
```

Schooling have a positive correlation with life expectancy.

```python
# What is the impact of Immunization coverage on life Expectancy?
#Hepatitis B (HepB) immunization coverage among 1-year-olds (%)

sns.regplot(x ='Hepatitis B',y ='Life expectancy ',data = data,marker = '+',color = 'g')
```

We can obsserve that there is no linear relation between Immunization coverage and life Expectancy

```python
# Lets see the relationship between life expectancy and expenditure
sns.regplot(x ='Total expenditure',y ='Life expectancy ',data = data,marker = '+',color = 'g')
```

For Most of the total Expenditure are less than 10 and for them life expextancy approx range of  45 to 85

#### Lets see the distribution of the continuous features

```python
for feature in num_features:
    dataset = data.copy()
    plt.hist(dataset[feature])
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.title(feature)
    plt.show()
```

We can observe the most of the features are not normally Distributed.So we have to convert it into the normal distribution in feature engineering
