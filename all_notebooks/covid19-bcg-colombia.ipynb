```python
df = pd.read_csv('../input/hackathon/BCG_country_data.csv')
df.head()
```

```python
colombia = df[(df['country_name']=='Colombia')].reset_index(drop=True)
colombia.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x="percentage_people_with_diabetes",data=colombia,palette="flag",edgecolor="black")
plt.xticks(rotation=45)
plt.yticks(rotation=45)
# changing the font size
sns.set(font_scale=1)
```

```python
fig = px.bar(df, 
             x='lockdown_start', y='lockdown_end', color_discrete_sequence=['crimson'],
             title='Covid19 in Colombia', text='population_per_km2')
fig.show()
```

```python
fig = px.bar(df, 
             x='year_of_gni_report', y='gni_per_capita_usd', color_discrete_sequence=['#27F1E7'],
             title='Covid19 in Colombia', text='population')
fig.show()
```

```python
fig = px.density_contour(df, x="lockdown_start", y="lockdown_end", color_discrete_sequence=['purple'], title='Covid19 in Colombia')
fig.show()
```

```python
fig = px.line(df, x="percentage_men_smokers", y="stroke_disease_disability", color_discrete_sequence=['darkseagreen'], 
              title="Covid19 in Colombia")
fig.show()
```

```python
fig = px.bar(df,
             y='internet_cctld',
             x='population_per_km2',
             orientation='h',
             color='population',
             title='Covid19 in Colombia',
             opacity=0.8,
             color_discrete_sequence=px.colors.diverging.Spectral,
             template='plotly_dark'
            )
fig.update_xaxes(range=[0,35])
fig.show()
```
