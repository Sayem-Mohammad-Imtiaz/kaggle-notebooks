```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("../input/startup_funding.csv", thousands=",", parse_dates=["Date"])
data.shape
```

```python
data.columns = data.columns.str.lower()
```

```python
data.head()
```

```python
data.describe(include=[np.object])
```

```python
def clean_dates(e):
    if e == "12/05.2015":
        e = "12/05/2015"
    if e == "13/04.2015":
        e = "13/04/2015"
    if e == "15/01.2015":
        e = "13/01/2015"
    if e == "22/01//2015":
        e = "22/01/2015"
    return e
```

```python
data["date_c"] = pd.to_datetime(data["date"].apply(clean_dates), format="%d/%m/%Y", errors="raise")
data["year"] = pd.DatetimeIndex(data["date_c"]).year
data["month"] = pd.DatetimeIndex(data["date_c"]).month
```

```python
 data.head()
```

```python
data[["citylocation", "amountinusd", "year"]] \
    .groupby(["year"], as_index=False) \
    .mean() \
    .round(2) \
    .plot(x="year", y="amountinusd", kind="bar", table=True) \
    .axes.get_xaxis().set_visible(False)
```

```python
data.loc[:, 'investment_type'] = data.investmenttype
data.loc[(data.investmenttype == "Crowd funding"), "investment_type"] = "Crowd Funding"
data.loc[(data.investmenttype == "PrivateEquity"), "investment_type"] = "Private Equity"
data.loc[(data.investmenttype == "SeedFunding"), "investment_type"] = "Seed Funding"
```

```python
 data.head()
```

```python
data[["investment_type", "year"]] \
    .groupby(["investment_type", "year"]) \
    .size() \
    .round(2) \
    .unstack(0) \
    .fillna(0) \
    .plot(kind="bar", stacked=True, table=True) \
    .axes.get_xaxis().set_visible(False)
```

```python
data["yearquarter"] = pd.PeriodIndex(data['date_c'], freq='Q').strftime('%Y-Q%q')
```

```python
data.head()
```

```python
data[["investment_type", "year", "yearquarter"]] \
    .groupby(["investment_type", "yearquarter"]) \
    .size() \
    .round(2) \
    .unstack(0) \
    .fillna(0) \
    .plot(kind="bar", stacked=True, table=True, figsize=(15,5)) \
    .axes.get_xaxis().set_visible(False)
```
