## Univariate analysis for Target value 0 and 1 making two different dataframes for both

```python
# Create a dataframe for Target value 0
target_0 = inp4[inp4['TARGET'] == 0]
```

```python
# Create a dataframe for Target value 0
target_1 = inp4[inp4['TARGET'] == 1]
```

```python
target_0.shape
```

```python
target_1.shape
```

```python
# Create Function to plot for compare the categorical variables with respect to Target_0 and Target_1 
def plot_target(item):
    plt.figure(figsize = (12,7))
    plt.subplot(1, 2, 1)
    sns.countplot(item, data = target_0, order = target_0[item].value_counts().index)
    plt.title('Graph of Non-Defaulters', fontsize=12)
    plt.xlabel(item)
    plt.xticks(rotation=90)
    plt.ylabel('Cases for non-defaulters')
    plt.subplot(1, 2, 2)
    sns.countplot(item, data=target_1, order = target_1[item].value_counts().index)
    plt.title('Graph of Defaulters', fontsize=12)
    plt.xlabel(item)
    plt.xticks(rotation=90)
    plt.ylabel('Cases for defaulters')
    plt.show()
```

```python
plot_target('NAME_CONTRACT_TYPE')
```

### Inference



Graph of NAME_CONTRACT_TYPE (Identification if loan is cash or revolving) vs TARGET (0 - Non-Defaulter, 1 - Defaulter)

According to the graph above, we can infer - percentage of population opting for revolving loans have lesser chance to default the loan payment

### Inference
