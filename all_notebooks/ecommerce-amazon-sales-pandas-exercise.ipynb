# Ecommerce Purchases Pandas Exercise

Exercise with fake data of Amazon Purchases (Ecommerce basic example). No real data was used (i.e. adresses, credit cards).

**Import Pandas and read csv file into a DataFrame.**

```python
import pandas as pd
ecom=pd.read_csv('../input/ecommerce-purchases-csv/Ecommerce Purchases.csv')
```

**Check the head of the DataFrame.**

```python
ecom.head(3)
```

**How many rows and columns are there? **

```python
#len(ecom.columns)
#len(ecom.index)
ecom.info()
```

**What is the average Purchase Price? **

```python
ecom['Purchase Price'].mean()
```

**What were the highest and lowest purchase prices? **

```python
ecom['Purchase Price'].max()
```

```python
ecom['Purchase Price'].min()
```

**How many people have English 'en' as their Language of choice on the website? **

```python
len(ecom[ecom['Language']=='en'])
# ecom[ecom['Language']=='en']['Language'].count()
```

**How many people have the job title of "Lawyer" ? **


```python
len(ecom[ecom['Job']=='Lawyer'].index)
```

**How many people made the purchase during the AM and how many people made the purchase during PM ? **

```python
ecom['AM or PM'].value_counts()
```

**What are the 5 most common Job Titles? **

```python
ecom['Job'].value_counts().head(5)
```

**Someone made a purchase that came from Lot: "90 WT" , what was the Purchase Price for this transaction? **

```python
ecom[ecom['Lot']=='90 WT']['Purchase Price']
```

**What is the email of the person with the following Credit Card Number: 4926535242672853 **

```python
ecom[ecom['Credit Card']==4926535242672853]['Email']
```

**How many people have American Express as their Credit Card Provider *and* made a purchase above $95 ?**

```python
len(ecom[(ecom['CC Provider']=='American Express') & (ecom['Purchase Price']>95)])
```

**How many people have a credit card that expires in 2025? **

```python
sum(ecom['CC Exp Date'].apply(lambda exp: exp[3:]=='25'))
```

**What are the top 7 most popular email providers/hosts (e.g. gmail.com, yahoo.com, etc...) **

```python
ecom['Email'].apply(lambda email: email.split('@')[1]).value_counts().head(7)
```
