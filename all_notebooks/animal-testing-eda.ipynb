## Terrace
That part is not perfect but I just want the show the specific animals because most time people react to these animals. However, other unfamiliar animals use commonly in that case (I know hamster is so familiar but I hear the Guinea Pigs first time).

```python
fig = go.Figure()
Data = [colD,colE,colC]
Domestic = ["Cats","Dogs","Rabbits" ,"Hamsters"]
size = []

for data in Data : 
    for ani in Domestic:
        size= data[ani].apply(lambda x :int(np.log(x+1))**2)
        fig.add_trace(go.Scatter(x=data.State,y=data[ani],name=ani,marker_size = size,hovertext = data["info"]))

fig.update_traces(mode="markers",marker=dict(sizemode="area",line_width=1))

fig.update_layout(title = "Domestic",xaxis=dict(title="State"),yaxis=dict(title="count"))


fig.show()
```

```python
fig = go.Figure()
Data = [colD,colE,colC]
Domestic = ["Pig","Guinea Pigs","Other Farm Animals","Sheep"]
size = []

for data in Data : 
    for ani in Domestic:
        size= data[ani].apply(lambda x :int(np.log(x+1))**2)
        fig.add_trace(go.Scatter(x=data.State,y=data[ani],name=ani,marker_size = size,hovertext = data["info"]))

fig.update_traces(mode="markers",marker=dict(sizemode="area",line_width=1))

fig.update_layout(title = "Farm animals",xaxis=dict(title="State"),yaxis=dict(title="count"))


fig.show()
```
