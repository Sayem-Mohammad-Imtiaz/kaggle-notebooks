```python
df = pd.read_csv('../input/housesalesprediction/kc_house_data.csv')
```

```python
df.head()
```

Target variable to predict: price

```python
Y=df['price']
```

```python
df.info()
```

```python
ColumnList=['bedrooms', 'sqft_living', 'sqft_lot', 'floors', 'yr_renovated', 'zipcode']
```

```python
X=df[ColumnList]
```

```python
X.head()
```

X - the columns to use for prediction

```python
from sklearn.model_selection import train_test_split
```

Split the data into training and testing

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

First mode class: linear regression

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
Ypred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
r2_score(Ypred,Y_test)
```

```python
mean_absolute_error(Ypred,Y_test)/Y_test.abs().mean()
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
```

```python
dt.fit(X_train,Y_train)
```

```python
Ypred=dt.predict(X_test)
```

```python
r2_score(Ypred,Y_test)
```

```python
mean_absolute_error(Ypred,Y_test)/Y_test.abs().mean()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(Y_test.index, Y_test, '-r', Y_test.index, Ypred, '-b' )
```

```python
from sklearn.svm import SVR
```

```python
svr=SVR()
svr.fit(X_train,Y_train)
YpredSVM=svr.predict(X_test)
```

```python
r2_score(YpredSVM,Y_test)
```

```python
mean_absolute_error(YpredSVM,Y_test)/Y_test.abs().mean()
```

```python
plt.plot(Y_test.index, Y_test, '-r', Y_test.index, YpredSVM, '-b' )
```

SVM does not predict well, let us try neural networks

```python
from sklearn.neural_network import MLPRegressor
```

```python
nr=MLPRegressor(hidden_layer_sizes=3*(10,10,10,10))
```

```python
nr.fit(X_train,Y_train)
```

```python
YpredNN=nr.predict(X_test)
```

```python
r2_score(YpredNN,Y_test)
```

```python
mean_squared_error(YpredNN,Y_test)
```

```python
mean_squared_error(YpredSVM,Y_test) > mean_squared_error(YpredNN,Y_test)
```

```python
plt.plot(Y_test.index, Y_test, '-r', Y_test.index, YpredNN, '-b' )
```

```python
mean_absolute_error(YpredNN,Y_test)/Y_test.abs().mean()
```
