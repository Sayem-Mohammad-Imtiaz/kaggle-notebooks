### Recorrido y actualización del dataframe

#### Valor máximo de una columna

```python
def valor_maximo_minimo_goles(lista):
    maximo = 0
    minimo = 100

    for a in lista:
        if (a > maximo):
            maximo = a
        elif(a<minimo):
            minimo = a
    
    return("{0} máximo de goles en casa\n{1} mínimo de goles en casa".format(maximo, minimo))
    
print(valor_maximo_minimo_goles(mydata['home_score']))
```

#### Actualizar variable cuantitativa

**a. Multiplicar goles de local y visitante por 2 constantes**

```python
def multiplicar_goles_local(goles):
    mg = goles * 2
    return mg

def multiplicar_goles_visitante(goles):
    mg = goles * 4
    return mg

mydata['home_score'] = mydata['home_score'].map(multiplicar_goles_local)
mydata['away_score'] = mydata['away_score'].map(multiplicar_goles_visitante)


mydata.head()
```

#### Colocar una variable qualitativa en mayúscula

**b. Transformando variables equipo visitante y local a mayúsculas**

```python
for a in mydata['home_team']:
    mydata['home_team'] = a.upper()


for a in mydata['away_team']:
    mydata['away_team'] = a.upper()

mydata.head()
```
