Reading in the Hipparcos database and reconstructing H-R diagram

```python
import pandas as pd
df_full = pd.read_csv("/kaggle/input/hipparcos-star-catalog/hipparcos-voidmain.csv")
df_full.columns
```

Only data with parallax greater than zero can be used so that you can find distance

```python
df = df_full[df_full["Plx"]>0.000001]
df = df.dropna(subset = ["B-V"])
df = df.sample(frac=.02)
```

```python
bmv = np.array(df["B-V"],dtype=float)
```

```python
m = np.array(df["Vmag"],dtype=float)
```

```python
p = np.array(df["Plx"],dtype=float)
```

without correcting for distance

```python
import matplotlib.pyplot as plt
plt.plot(bmv,-m,'.',markersize=0.05)
plt.show()
```

Use parallax to find distance

```python
d = 1/p*1000
```

```python
import numpy as np
```

Correct for absolute magnitude

```python
M=m-5*np.log10(d/10)
```

```python
plt.plot(bmv,-M,'.',markersize=1.0)
plt.show()
```

```python
M = M/10
dataToCluster = np.transpose(np.array([bmv,M]))
dataToCluster
```

```python
np.max(bmv)
```

```python
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=.07,min_samples=20)
solution = db.fit(dataToCluster)
np.max(solution.labels_)
```

```python
from sklearn.cluster import OPTICS
db=OPTICS(min_samples=10)
solution = db.fit(dataToCluster)
np.max(solution.labels_)
```

```python
from sklearn.cluster import AgglomerativeClustering
db=AgglomerativeClustering(n_clusters=None,distance_threshold=20)
solution = db.fit(dataToCluster)
np.max(solution.labels_)
```

```python
import plotly.express as px
fig = px.scatter(x= bmv,y=-M,color=solution.labels_)
fig.update_traces(marker=dict(size=5))
fig.show()
```
