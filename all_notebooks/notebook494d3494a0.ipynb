```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
data=pd.read_csv('../input/logistic-regression/Social_Network_Ads.csv')
data.head()
```

```python
data['Gender']=data['Gender'].map({'Male':2,'Female':3})
data.head()
```

```python
data.isna().sum()
```

```python
positive=data[data['Purchased'].isin([1])]
negative=data[data['Purchased'].isin([0])]

fig,ax=plt.subplots(figsize=(15,10))



ax.scatter(positive['Gender'],positive['Age'],s=50,c='b',marker='o',label='Purchased')
ax.scatter(negative['Gender'],negative['Age'],s=50,c='r',marker='x',label='Not Purchased')
```

```python

fig,ax=plt.subplots(figsize=(15,10))

ax.scatter(positive['EstimatedSalary'],positive['Gender'],s=50,c='b',marker='o',label='Purchased')
ax.scatter(negative['EstimatedSalary'],negative['Gender'],s=50,c='r',marker='x',label='Not Purchased')
```

```python
fig,ax=plt.subplots(figsize=(15,10))

ax.scatter(positive['EstimatedSalary'],positive['Age'],s=50,c='b',marker='o',label='Purchased')
ax.scatter(negative['EstimatedSalary'],negative['Age'],s=50,c='r',marker='x',label='Not Purchased')
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X=data.drop(['Purchased','Gender','User ID'],axis=1).values
X=scaler.fit_transform(X)
y=data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
logRegressor=LogisticRegression()
logRegressor.fit(X_train,y_train)
```

```python
pred=logRegressor.predict(X_test)
pred=np.reshape(pred,(80,1))
y_test=np.reshape(y_test,(80,1))
logRegressor.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix,accuracy_score
print(classification_report(y_test,pred))
print ("Confusion matrix : \n",confusion_matrix(y_test,pred))
print ("Accuracy : \n",accuracy_score(y_test,pred))
```
