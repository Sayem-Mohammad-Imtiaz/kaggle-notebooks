```python
df=pd.read_csv('../input/coursera-course-dataset/coursea_data.csv',thousands=',')
```

```python
#change datatype
df['course_rating'].astype('int64')
```

```python
#change k to 1000 and change datatype
df['course_students_enrolled']=df['course_students_enrolled'].replace({'k': '*1e3', 'm': '*1e6'}, regex=True).map(pd.eval).astype(int)
df['course_students_enrolled'][:5]
```

```python
df.drop_duplicates()
```

```python
#Checking null
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df.sort_values(by='course_rating',inplace=False,ascending=False)
```

```python
#Top 10 organizations based on courses offered
Ten_org = pd.DataFrame(df['course_organization'].value_counts()).head(10)

sns.barplot(x = Ten_org.index, y = Ten_org['course_organization'],palette='icefire')

labels =Ten_org.index.tolist()
plt.gcf().set_size_inches(15,7)

plt.title('Organization vs courses offered', fontsize = 20)
plt.xlabel('Organization', fontsize = 15)
plt.ylabel('courses offered', fontsize = 15)

plt.xticks(np.arange(10) ,labels =labels,rotation = 90)
plt.show()
```

```python
#Top 10 organizations based on popularity or student enrollment
find_pop_org=df[['course_organization','course_students_enrolled']]
Ten_pop_org = find_pop_org.nlargest(10,['course_students_enrolled'])
sns.barplot(y = Ten_pop_org['course_organization'], x = Ten_pop_org['course_students_enrolled'],orient='h',palette='icefire')
plt.xticks(rotation = 90)
```

```python
Ten_pop_org
```

```python
#Top 10 courses based on rating
find_course=df[['course_title','course_rating']]
Ten_course = find_course.nlargest(10,['course_rating'])
sns.barplot(y = Ten_course['course_title'], x = Ten_course['course_rating'],orient='h',palette='rocket')
plt.xticks(rotation = 90)
```

```python
#Top 10 courses based on popularity or student enrollment
find_pop_course=df[['course_title','course_students_enrolled']]
Ten_pop_course = find_pop_course.nlargest(10,['course_students_enrolled'])
sns.barplot(x = Ten_pop_course['course_students_enrolled'], y = Ten_pop_course['course_title'],orient='h',palette='rocket')
plt.xticks(rotation = 90)
```
