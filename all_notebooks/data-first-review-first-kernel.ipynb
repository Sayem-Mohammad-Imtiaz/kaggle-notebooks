# Basic Data Visualization

```python
f,ax = plt.subplots(figsize = (10,10)) # We are creating table for new heatmap
data = df_19.loc[df_19["Social support"] >= 1.350,:] # We are filtering this data because We look correlations to it
corr = np.triu(data.corr()) # This part is output without correlations
sns.heatmap(data.corr(),annot = True,mask = corr) # And look at the heatmap
```

```python
df_19["Healthy life expectancy"].plot(kind = "line", color = "g", label = "Health (Life Expectancy)", linewidth = 5, grid = True, alpha = 0.5, linestyle = ":")
df_19["GDP per capita"].plot(color = "r", label = "Economy (GDP per Capita)", grid = True, alpha = 0.5, linestyle = "-.")
plt.legend(loc = "upper right")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line Plot") # This part is visualizations part and we comparison data
# Gives better results in terms of time in a line plot but now we hadn't time variables and we look counrty rank
```

```python
sns.scatterplot(x = df_19["Freedom to make life choices"], y = df_19["GDP per capita"],hue = df_19["Score"]);
# We look these variables have how correlation. They have positive correlation
# This plot best feature is the graph showing the correlation between 2 variables.
```

```python
sns.pairplot(df_19); # These plots are scatter and histogram plots for all variables
```

```python
df_19["Freedom to make life choices"].plot(kind = "hist", bins = 50, figsize = (10,10)) 
# A histogram is a graph showing the distribution within a variable.
plt.show()
```
