```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
from sklearn.neighbors import KNeighborsRegressor
import numpy as np
```

```python
import pandas as pd
# load dataset
col_names = ['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
pima = pd.read_csv("../input/pima-indians-diabetes-database/diabetes.csv")
```

```python
x = pima.iloc[:,:-1]# Features
y = pima.iloc[:, 8]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
# prediction without scaling
logreg = LogisticRegression()

# fit the model with data
logreg.fit(x_train,y_train)

y_pred=logreg.predict(x_test)
```

```python
from sklearn import metrics
print("Accuracy/normal:",metrics.accuracy_score(y_test, y_pred))
print("Precision:",metrics.precision_score(y_test, y_pred))
print("Recall:",metrics.recall_score(y_test, y_pred))
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler

# fit scaler on training data
norm = MinMaxScaler().fit(x_train)
# transform training data
x_train_norm = norm.transform(x_train)
# transform testing dataabs
x_test_norm = norm.transform(x_test)
```

```python
# fit scaler on training data
stand = StandardScaler().fit(x_train)
# transform training data
x_train_stand = stand.transform(x_train)
# transform testing dataabs
x_test_stand = stand.transform(x_test)
```

```python
# predict normalized data
logreg1 = LogisticRegression()
# fit the model with data
logreg1.fit(x_train_norm,y_train)
y_pred1=logreg1.predict(x_test_norm)
```

```python
print("Accuracy/normalized:",metrics.accuracy_score(y_test, y_pred1))
print("Precision:",metrics.precision_score(y_test, y_pred1))
print("Recall:",metrics.recall_score(y_test, y_pred1))
```

```python
# predict standardized data
logreg2 = LogisticRegression()
# fit the model with data
logreg2.fit(x_train_stand,y_train)
y_pred2=logreg2.predict(x_test_stand)
```

```python
print("Accuracy/standardized:",metrics.accuracy_score(y_test, y_pred2))
print("Precision:",metrics.precision_score(y_test, y_pred2))
print("Recall:",metrics.recall_score(y_test, y_pred2))
```

```python


knn = KNeighborsRegressor(n_neighbors=7)
rmse = []
accuracy = []
# raw, normalized and standardized training and testing data
trainX = [x_train, x_train_norm, x_train_stand]
testX = [x_test, x_test_norm, x_test_stand]

# model fitting and measuring RMSE
for i in range(len(trainX)):
    
    # fit
    knn.fit(trainX[i],y_train)
    # predict
    pred = knn.predict(testX[i])
    # RMSE
    rmse.append(np.sqrt(mean_squared_error(y_test,pred)))
    accuracy.append(metrics.accuracy_score(y_test,pred.round()))

# visualizing the result
df_knn = pd.DataFrame({'RMSE':rmse,'ACCURACY':accuracy},index=['Original','Normalized','Standardized'])
df_knn
```
