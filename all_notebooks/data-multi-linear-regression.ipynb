importing library
*  NumPy is a package in Python used for Scientific Computing. The ndarray (NumPy Array) is a multidimensional array used to store values of same datatype
* matplotlib.pyplot for data visualization 
* pandas for data manipulation  


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

importing dataset to x and y

```python
dataset = pd.read_csv('../input/50-startups/50_Startups.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values
```

convert categorical data, or text data, into numbers, which our predictive models can better understand.

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(
    [('one_hot_encoder', OneHotEncoder(), [3])],    #number 3 means which Column
    remainder='passthrough'                         # Leave the rest of the columns untouched
)
x = np.array(ct.fit_transform(x), dtype=np.float)
```

```python
x=x[:,1:] # Avoiding the Dummy Variable Trap
```

Splitting the dataset into the Training set and Test set
* note: early its was in sklearn.cross_validation for new version it will be in sklearn.model_selection

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_teast=train_test_split(x,y,test_size=0.2,random_state=0)
```

no need for feature scaling here

simple linear regression traning

```python
#traning
from sklearn.linear_model import LinearRegression
regg=LinearRegression()
regg.fit(x_train,y_train)
```

simple linear regression prediction

```python
y_pred=regg.predict(x_test)
```

building optimal model using backward elimination

```python
import statsmodels.api as sm
x=np.append(arr=np.ones((50,1)).astype(int),values=x,axis=1)
x_opt=x[:,[0,1,2,3,4,5]]
regg_ols=sm.OLS(endog=y,exog=x_opt).fit()
regg_ols.summary()
```

```python
x_opt=x[:,[0,1,3,4,5]]
regg_ols=sm.OLS(endog=y,exog=x_opt).fit()
regg_ols.summary()
```

```python
x_opt=x[:,[0,3,4,5]]
regg_ols=sm.OLS(endog=y,exog=x_opt).fit()
regg_ols.summary()
```

```python
x_opt=x[:,[0,3,5]]
regg_ols=sm.OLS(endog=y,exog=x_opt).fit()
regg_ols.summary()
```

```python
x_opt=x[:,[0,3]]
regg_ols=sm.OLS(endog=y,exog=x_opt).fit()
regg_ols.summary()
```
