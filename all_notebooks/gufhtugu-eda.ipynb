```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.io as pio
import plotly.graph_objects as go
```

```python
ds = pd.read_csv("../input/gufhtugu-publications-dataset-challenge/GP Orders - 5.csv", sep=",", encoding="utf_8")
```

```python
ds.head(10)
```

```python
ds.describe()
```

```python
ds.info()
```

```python
ds.isnull().sum()
```

```python
ds.dropna(inplace=True)
print("\n-------------------After Droping null values----------------------\n")
ds.isnull().sum()
```

```python
ds.columns
```

```python
columns = ds.columns.str.replace(' ', '_')
```

```python
ds.columns = columns
```

```python
ds.head()
```

```python
ds['Order_Date_&_Time'] = pd.to_datetime(ds['Order_Date_&_Time'])
```

Master Calendar


```python
ds['Date'] = pd.to_datetime(ds['Order_Date_&_Time']).dt.date
ds['Day'] = ds['Order_Date_&_Time'].dt.day_name()
ds['Month'] = ds['Order_Date_&_Time'].dt.month_name() 
ds['Quarter'] = ds['Order_Date_&_Time'].dt.quarter 
ds['Year'] = ds['Order_Date_&_Time'].dt.year
```

```python
ds.head(1)
```

```python
ds_order_status_count = ds.Order_Status.value_counts().to_frame('count')
```

```python
ds_count = ds.Order_Number.count()
```

# Visualization

No of Orders by Status

```python
fig=px.pie(ds,names="Order_Status")
fig.update_traces(textinfo='value')
fig.show()
```

# Order Status by Payment Methods 

```python
plt.figure(figsize=(15,8))
sns.countplot(x="Order_Status",data=ds,hue="Payment_Method")

plt.xlabel("Order_Status")
plt.ylabel("Order_Numbers")
```

#  Top 10 Best Selling Books

```python
best_selling_books=ds[ds["Order_Status"]=="Completed"]["Book_Name"].value_counts().head(10).reset_index()
fig=px.bar(best_selling_books,x="index",y="Book_Name",color="index",title="10 Best Selling Books")
fig.update_layout(xaxis_title="Book_Name",yaxis_title="Orders")
fig.show()
```

# No of Orders by Payment Method

```python
fig=px.pie(ds,names="Payment_Method")
fig.update_traces(textinfo='value')
fig.show()
```

# Visualization by Year, Quarter and Months

```python
fig = px.pie(ds,names="Year",labels="Year",title="No of Orders by Year")  
fig.update_traces(textinfo='value')
fig.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x="Year",data=ds,hue="Quarter")

plt.xlabel("Year")
plt.ylabel("Order_Numbers")
```

```python
fig = px.pie(ds,names="Month",labels="Month",title="No of Orders by Month")  
fig.update_traces(textinfo='value')
fig.show()
```

# No of Orders by Month Year

```python
plt.figure(figsize=(15,8))
sns.countplot(x="Year",data=ds,hue="Month")

plt.xlabel("Year")
plt.ylabel("Order_Number")
```
