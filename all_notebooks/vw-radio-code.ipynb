```python
#check for duplication
df=df.drop_duplicates( ignore_index=True)
```

```python
#Preapre data according
#vin number should be > 17digits
count=0
total=75
for i in df['VIN']:
    if len(i)==17:
        count+=1
print(count)
```

```python
#Radio codes must be 4 digits

df=df[df['Radio Code']  >= 1000]
```

```python
#Selection of feature variable
X=df[['VIN','Radio Serial Number']]
Y=df[['Radio Code']]
```

```python
#Spliting the traing and testing data
from sklearn.model_selection import train_test_split
```

```python

xtrain,xtest,ytrain,ytest=train_test_split(X,Y, test_size=.25, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
#Apply Kfold Cross Validation for choosing best model.
models=[LogisticRegression(),SVC(),DecisionTreeClassifier(),RandomForestClassifier(),XGBClassifier(),KNeighborsClassifier()]

from sklearn.model_selection import KFold,cross_val_score
def train(model):
  kfold=KFold(n_splits=20,random_state=101,shuffle=True)
  pred=cross_val_score(model,X,Y,cv=kfold)
  cv_score=pred.mean()
  print('Model:',model)
  print('CV-score:',cv_score)

for model in models:
  train(model)
```
