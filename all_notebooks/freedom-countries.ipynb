```python
#Codes by Tawej https://www.kaggle.com/tawejssh/covid-19

#Let's visualise freedom
freedom = tenNF.groupby('fiw_survey_edition').sum()[['number_of_f_countries','number_of_pf_countries','number_of_nf_countries']]
#evolution['Expiration Rate'] = (evolution['Expired'] / evolution['Cumulative']) * 100
#evolution['Discharging Rate'] = (evolution['Discharged'] / evolution['Cumulative']) * 100
freedom.head()
```

```python
#Codes by Tawej https://www.kaggle.com/tawejssh/covid-19

plt.figure(figsize=(20,7))
plt.plot(freedom['number_of_f_countries'], label='number_of_f_countries')
plt.plot(freedom['number_of_pf_countries'], label='number_of_pf_countries')
plt.plot(freedom['number_of_nf_countries'], label='number_of_nf_countries')
plt.legend()
#plt.grid()
plt.title('Countries Freedom last 10 Years - FIW')
plt.xticks(freedom.index,rotation=45)
plt.xlabel('fiw_survey_edition')
plt.ylabel('Count')
plt.show()
```

```python
#Codes by Tawej https://www.kaggle.com/tawejssh/covid-19

#What aboutfreedom in Not Free countries rate ?
plt.figure(figsize=(20,7))
plt.plot(freedom['number_of_nf_countries'], label='Number of Not free Countries')
plt.legend()
#plt.grid()
plt.title('Number of Not free Countries last 10 years')
plt.xticks(freedom.index,rotation=45)
plt.ylabel('Rate %')
plt.show()
```

```python
#Codes by Tawej https://www.kaggle.com/tawejssh/covid-19

#This is another way of visualizing freedom: plotting the increase freedom on the last 10 years
diff_freedom = freedom.diff().iloc[1:]
plt.figure(figsize=(20,7))
plt.plot(diff_freedom['number_of_nf_countries'], label='Number of Not Free  Countries')
plt.legend()
plt.grid()
plt.title('Number of Not free Countries last 10 years')
plt.xticks(freedom.index,rotation=45)
plt.ylabel('Rate %')
plt.show()
```

```python
#Code by Debadri Dutta https://www.kaggle.com/duttadebadri/why-suicides-are-a-major-concern/notebook

df.fillna(df.mean(),inplace=True)
df_free=pd.DataFrame(df.groupby(['total_countries','fiw_survey_edition'])['number_of_f_countries'].sum().reset_index())
df_free.head()
```

```python
#Code by Debadri Dutta https://www.kaggle.com/duttadebadri/why-suicides-are-a-major-concern/notebook

df_fiw=df_free[['fiw_survey_edition','number_of_f_countries']]
df_fiw.index=df_fiw['fiw_survey_edition']


df_fiw['number_of_f_countries'].plot(figsize=(15,6), color="green")
plt.xlabel('FIW Survey Edition')
plt.ylabel('Number of Free Countries')
plt.title("Number of Free Countries")
plt.show()
```
