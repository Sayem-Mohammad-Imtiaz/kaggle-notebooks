```python
# USER DEFINED FUNCTION
def example():
    t = (1,2,3)
    return t
a,b,c = example()
print(a,b,c)
```

```python
# scope
x = 3
def f():
    x = 5
    return x
print(x)      # x = 3 global scope
print(f())    # x = 5 local scope
```

```python
# What if there is no local scope
x = 3
def f():
    y = 9*x        # there is no local scope x
    return y
print(f())         # it uses global scope x
# First local scopesearched, then global scope searched, if two of them cannot be found lastly built in scope searched.
```

```python
#nested function 
def example():
    def other():
        x = 3
        y = 5
        z = x + y
        return z
    return other()**2
print(example())    
```

```python
# default arguments
def f(a, b = 2, c = 3):
    y = a + b + c
    return y
print(f(5))
# what if we want to change default arguments
print(f(5,4,3)) 
```

```python
# flexible arguments *args
def f(*args):
    for i in args:
        print(i)
f(3)
print("")
f(1,2,3,4)
```

```python
# flexible arguments **kwargs that is dictionary
def f(**kwargs):
    for key, value in kwargs.items():               
        print(key, " ", value)
f(country = 'turkey', capital = 'ankara', population = 6000000)
```

```python
# lambda function
square = lambda x: x**2     # where x is name of argument
print(square(5))
total = lambda x,y,z: x+y+z   # where x,y,z are names of arguments
print(total(1,2,3))
```

```python
#ANONYMOUS FUNCTION
number_list = [1,2,3]
y = map(lambda x:x**2,number_list)
print(list(y))
```

```python
# zip example
list1 = [1,2,3,4]
list2 = [5,6,7,8]
z = zip(list1,list2)
print(z)
z_list = list(z)
print(z_list)
```

```python
un_zip = zip(*z_list)
un_list1,un_list2 = list(un_zip) # unzip returns tuble
print(un_list1)
print(un_list2)
print(type(un_list2))
print(type(list(un_list1)))
```

```python
# Conditionals on iterable
num1 = [5,10,15,3,8,9]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)

a = ["small" if x < 9 else "equal" if x==9 else "big" for x in num1] 
print(a)
```
