# Load Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

# Read Data

```python
df = pd.read_csv('../input/50-startups/50_Startups.csv')
df.head()
```

# Check if null

```python
df.isnull().sum()
```

# Managing Categorical Values

```python
dummy = pd.get_dummies(df["State"])
df = pd.concat([dummy, df], axis=1)
df.drop(['State'], inplace=True, axis=1)
df.head()
```

# Correlation

```python
df.corr()
```

# Split x and y values

```python
x_copy = df.copy()

x_copy.drop(['California', 'Florida', 'New York','Administration'], inplace=True, axis=1)
x_copy.head()
```

```python
X = x_copy.iloc[:, 0:2].values
y = x_copy.iloc[:, -1:].values
```

# Splliting the data

```python
X_train, X_test, y_train, Y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

# Training the data

```python
regressor = LinearRegression()
# Training the model
regressor.fit(X_train, y_train)
```

# Predict the data

```python
y_pred = regressor.predict(X_test)
```

```python
print("="*40)
print("Real Value ---------> Predicted Value")
print("="*40)
print("\n")

for item in range(len(Y_test)):
    print(str(Y_test[item]) + " --------> " + str(y_pred[item]))
```

# Calulating r2 score

```python
r2_score(Y_test, y_pred)
```
