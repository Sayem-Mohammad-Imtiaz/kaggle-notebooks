```python
# This code creates a method that takes a variable name as argument and return the feature that is highly correlated to the given variable
# Also this makes sure that the returned value is not the same variable name that is passed as input

import numpy as np # linear algebra
import pandas as pd # data processing

df = pd.read_csv('../input/craigslist-carstrucks-data/vehicles.csv') # Use your dataset here
df.head(5)

def get_max_correlated_column(i):

    #Creating Correlation matrix
    cor_matrix = df.corr().abs()

    # Convert Matrix to Dataframe
    df1 = cor_matrix.stack().reset_index()
    df1.columns = ['ColX','ColY','correlation']

    # get names of indexes for which ColY and ColX are same
    index_names = df1[ df1['ColY'] == df1['ColX']].index

    # drop these row indexes from dataFrame
    df1.drop(index_names, inplace = True)

    # Rank ColY based on correlation for each ColX
    df1['rank_corr'] = df1.sort_values(['correlation'], ascending=False).groupby(['ColX']).cumcount() + 1

    # Keep only rows that has highest correlation i.e Rank_corr = 1
    highest_corr_df = df1
    highest_corr_df.drop(highest_corr_df[ highest_corr_df['rank_corr'] > 1].index, inplace = True)

    # Highest correlated variable to the given variable 
    return highest_corr_df.loc[highest_corr_df['ColX'] == i, 'ColY'].iloc[0]
    
```
