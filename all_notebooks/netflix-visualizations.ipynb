# Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'/kaggle/input/netflix-shows/netflix_titles.csv')
```

```python
df.shape #dimension
```

```python
df.describe(include='all') #statistical sumaary
```

```python
df.isnull().sum()# checking for missing values
```

all the columns which has missing values we'll hae to manually put values for them

```python
df.head()
```

```python
df['type'].value_counts().plot(kind='pie',startangle=90,shadow=True,explode=[0,0.07],colors=['black','red'],figsize=(7,7),autopct='%1.1f%%',textprops={'color':"w"})
plt.title('Ratio of Movie vs TV show')
plt.show()
```

```python
df['release_year'].value_counts().sort_values(ascending=False).plot(kind='bar',figsize=(20,8),color='red')
plt.title('Count of movie and tv shows released in each year')
plt.xlabel('year')
plt.ylabel('count')
plt.show()
```

We can see that the maximum number of shows and movies are released in 2018

```python
from wordcloud import WordCloud
wc = WordCloud(background_color='black',max_words=500)
fig=plt.figure(figsize=(10,6))
wc.generate(str(df['description']))
plt.imshow(wc, interpolation="bilinear")
plt.title("Words used to describe the tv show or movies")
plt.axis('off')
plt.show()
```

Documentary, devastating, island, inhabit and recruit are some of the top words used to describe the movies and tv shows

```python
df['rating'].value_counts().sort_values(ascending=False).plot(kind='bar',figsize=(10,6),color='red')
plt.title('Count of movie and tv shows released in each different ratings')
plt.xlabel('rating')
plt.ylabel('count')
plt.show()
```

TV-MA(viewed by mature people) is the top rating followed by TV-14(for childrens over age 14) and TV-PG(unsuitable for children)

```python
df['country'].value_counts().sort_values(ascending=False).head(10).plot(kind='bar',figsize=(10,6),color='red')
plt.title('Countries producing maximum content')
plt.xlabel('Countries')
plt.ylabel('count')
plt.show()
```

US created the maximum amount of content followed by India

```python
df.head()
```

```python
df['director'].value_counts().sort_values(ascending=False).head(10).plot(kind='bar',figsize=(10,6),color='red')
plt.title("Top 10 directors who has produced the maximum amount of content")
plt.xlabel('Genre')
plt.ylabel('count')
plt.show()
```
