Hi, this notebook is for find out trend of Meditation App : Calm and Meditopia.

# 1. Loading Library

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

# 2. Loading data & Adding app name & country

I could not find which country source of Meditopia app. I assumed 'all' on country column.

```python
appstore_meditopia = pd.read_csv("/kaggle/input/app-store-meditation-app-reviews/appstore_meditopia.csv")
appstore_meditopia = appstore_meditopia.drop(['developerResponse.body'], axis=1)
appstore_meditopia["app_name"] = "meditopia"
appstore_meditopia["country"] = "all"


appstore_calm_us = pd.read_csv("/kaggle/input/app-store-meditation-app-reviews/appstore_calm_us.csv")
appstore_calm_us["app_name"] = "calm"
appstore_calm_us["country"] = "US"

appstore_calm_tr = pd.read_csv("/kaggle/input/app-store-meditation-app-reviews/appstore_calm_tr.csv")
appstore_calm_tr["app_name"] = "calm"
appstore_calm_tr["country"] = "Turkey"

appstore_total = pd.concat([appstore_meditopia, appstore_calm_us, appstore_calm_tr])
appstore_total["date"] = pd.to_datetime(appstore_total["date"]) # change Dtype from object to datetime64
appstore_total["year_month"] = appstore_total['date'].dt.strftime('%Y-%m')
```

# 3. Line Graph by Apps and by Countries

Pivot ratings value by App and YYYYMM format date

```python
ratings = pd.pivot_table(appstore_total, values='rating', index=['app_name', 'year_month'], aggfunc=np.mean)
ratings = ratings.sort_values(by=['app_name', 'year_month'], axis=0)
```

## Average Ratings by Apps

Meditopia has higher trend than Calm, but has greater volatility.

```python
plt.figure(figsize=(20,6))
plt.xticks(rotation = - 45 , fontsize=9)
sns.lineplot(data=ratings, x="year_month", y="rating",  hue="app_name")
```

Pivot ratings value by Country and YYYYMM format date (Calm only)

```python
ratings_calm = appstore_total[appstore_total['app_name'] == 'calm']
ratings_calm = pd.pivot_table(ratings_calm, values='rating', index=['country', 'year_month'], aggfunc=np.mean)
ratings_calm = ratings_calm.sort_values(by=['country', 'year_month'], axis=0)
```

## Average Ratings by Country

US users have liked Calm app. more than Turkey users.

```python
plt.figure(figsize=(20,6))
plt.xticks(rotation = - 45 , fontsize=9)
sns.lineplot(data=ratings_calm, x="year_month", y="rating",  hue="country")
```
