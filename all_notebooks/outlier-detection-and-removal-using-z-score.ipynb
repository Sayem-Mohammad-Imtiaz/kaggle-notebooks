```python
# Required Libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Loading data into Data Frame
df = pd.read_csv('/kaggle/input/gender-and-height-dataset/gender-height.csv')
```

```python
df.head()
```

```python
# Shape of Data Set
df.shape
```

```python
# Plotting Histogram on Heights
plt.hist(df.Height, bins=20, rwidth = 0.8)
plt.xlabel('Heights(inches)')
plt.ylabel('Counts')
plt.show()
```

```python
# Z score
df['Zscore'] = (df.Height - df.Height.mean())/df.Height.std()
```

```python
# Head of Data set
df.head()
```

```python
# Value above and below 3 SD
df[(df.Zscore > 3) | (df.Zscore < -3)]
```

```python
# Removing above Outliers from data
df_new = df[(df.Zscore < 3) & (df.Zscore > -3)]
```

```python
# Head of New Cleaned data
df_new.head()
```

```python
# Number of Outliers removed
print('Number of Outliers Removed is ',(df.shape[0] - df_new.shape[0]))
```
