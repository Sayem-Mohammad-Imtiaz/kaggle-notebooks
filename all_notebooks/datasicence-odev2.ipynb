```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("../input/pokemon-challenge/pokemon.csv")
```

```python
#Global scope dediğimiz her yerde kullanılabilen variabler, local scope sadece function ve method içinde kullanılanlar, build in scope ise len, print gibi
#hazır metodlara denir ve bunlar variable olarak kullanılmaz. Eğer local değişken function içinde yoksa aynı isimde global değişkene bakar

x = 5  #global değişken

def f():
    y = 2*x  #x ve y local değişken
    return y

print(f())   #burada f çağırılırken x aranıyor ve x değeri local olarak verilmediği için bu sefer global değişken arıyor ve 5i bulduğu için fonksiyon çalışıyor
```

```python
#nested function iç içe fonksiyon demektir
def calculate(a, b):     #burada, print kısmından a=2, b=4 atadık
    def multiply(a, b):   #çarpma işlemi yaptık
        c = a*b
        return c
    def sumation(a, b):      #toplama işlemi yaptık
        c = a+b
        return c
    return multiply(a,b)+sumation(a,b)      #return ettik
print(calculate(2,4))    #ve yazdırdık
```

```python
#default argument demek fonksiyon içinde başka bir atama yapılmazsa kullanılan değerlerdir
def f(a, b=1, c=2):   #b ve c default değerler almıştır
    y = a+b+c
    return y
print(f(5))  #burada a=5 olur b ve c default değerlerini alır
print(f(1,5,6))  #burada 1 a, 5 b, 6 c olmuştur, default değerler kullanılmaz.
```

```python
#flexible argument yani args yaparak istediğimiz kadar veri ekleyebiliyoruz
def f(*args):
    for i in args:
        print(i)

print(f(1,2,3,4,5))
```

```python
#list compherension yani listeden liste türetmektir
num1 = [5,10,15]
num2 = [i+1 for i in num1]   #num1 elemanlarını 1er arttırıp num2ye atadık
num3 = [i**2 if i<10 else i+1 if i==10 else i*20 for i in num1] #yani eğer 10dan küçükse karesi, eşitse 1 fazlası, değilse 20 katı yapıyoruz num1 elemanlarıyla

print(num2)
print(num3)
```

```python
threshold = sum(data.Speed)/len(data.Speed)
data["speed_level"] =["high" if i>threshold else "low"  for i in data.Speed]

data.loc[:10,["speed_level","Speed"]]
```

```python
#Attack ve Defens özelliklerine göre conditional list comprehension kullanacak ve zip metoduyla pokemonların genel olarak güçlü olup olmadığını histogramla gösterme
data["power"] = ["powerful" if i>75 and j>75 else "normal" if i<=75 and j<=75 and i>50 and j>50 else "weak" for i,j in zip(data.Attack, data.Defense)]
filtre = (data.power == "powerful")
print(filtre)
plt.hist(data.power, bins=50)
plt.show()
```
