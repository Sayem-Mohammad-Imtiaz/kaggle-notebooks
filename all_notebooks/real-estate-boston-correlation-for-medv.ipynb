```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import pylab as pl
import matplotlib.pyplot as plt
%matplotlib inline
data=pd.read_csv("../input/real-estate-dataset/data.csv")
```

```python
data.head()
```

```python
data.isnull().any()
```

```python
data.fillna(data.mean(), inplace=True)
```

```python
data.isnull().any()
```

```python
data.corr() 
```

```python
plt.subplots(figsize=(12,7))
sns.heatmap(data.corr(), annot=True, cmap='YlGnBu', linewidths=1 );
```

Correlation between CRIM and MEDV

```python
plt.scatter(data.CRIM, data.MEDV, color='green', alpha=0.8)
plt.xlabel("Criminal")
plt.ylabel("Value")
plt.show()
```

```python
house = np.random.rand(len(data)) < 0.8
train = data[house]
test = data[~house]
```

```python
plt.scatter(train.CRIM, train.MEDV,  color='blue')
plt.xlabel("Criminal")
plt.ylabel("Value")
plt.show()

from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['CRIM']])
train_y = np.asanyarray(train[['MEDV']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(train.CRIM, train.MEDV,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Criminal")
plt.ylabel("Value")

from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['CRIM']])
test_y = np.asanyarray(test[['MEDV']])
test_y_hat = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_hat - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_hat , test_y) )
```

Correlation between RM and MEDV

```python
plt.scatter(train.RM, train.MEDV,  color='blue')
plt.xlabel("Room")
plt.ylabel("Value")
plt.show()

regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['RM']])
train_y = np.asanyarray(train[['MEDV']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(train.RM, train.MEDV,  color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Room")
plt.ylabel("Value")


test_x = np.asanyarray(test[['RM']])
test_y = np.asanyarray(test[['MEDV']])
test_y_hat = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_hat - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_hat , test_y) )
```

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs 
from sklearn.preprocessing import StandardScaler
dt = data[['CRIM','RM', 'MEDV']]
X = dt.values[:,1:]
X = np.nan_to_num(X)
Clus_dataSet = StandardScaler().fit_transform(X)
Clus_dataSet
```

```python
clusterNum = 3
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
labels = k_means.labels_
print(labels)
```

```python
dt["data_km"] = labels
dt.head(5)
```

```python
dt.groupby('data_km').mean()
```
