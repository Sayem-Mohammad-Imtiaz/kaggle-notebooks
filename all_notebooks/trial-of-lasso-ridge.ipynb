```python
df=pd.read_csv('/kaggle/input/student-study-hour-v2/Student Study Hour V2.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
x=df.Hours.to_frame()
y=df.Scores
plt.scatter(x,y)
```

```python
lr=LinearRegression()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=123)
```

1.Linear Regression

```python
lr.fit(x_train,y_train)
print(lr.coef_)
print(lr.intercept_)
```

```python
lr.fit(x_train,y_train)
print(lr.coef_)
print(lr.intercept_)

y_pred=lr.predict(x_test)

print(y_pred-y_test)
print('MAE:',mean_absolute_error(y_pred,y_test ))
print('MSE:',mean_squared_error(y_pred, y_test))
print('RMSE:',np.sqrt(mean_squared_error(y_pred, y_test)))
print('R2:',r2_score(y_pred, y_test))
```

2. Lasso

```python
lasso=Lasso(alpha=0.001,normalize=True)
lasso.fit(x_train,y_train)

print(lasso.coef_)
print(lasso.intercept_)

y_pred2=lasso.predict(x_test)

print(y_pred2-y_test)
print('MAE:',mean_absolute_error(y_pred2,y_test ))
print('MSE:',mean_squared_error(y_pred2, y_test))
print('RMSE:',np.sqrt(mean_squared_error(y_pred2, y_test)))
print('R2:',r2_score(y_pred2, y_test))
```

3.Ridge

```python
ridge=Ridge(alpha=0.01,normalize=True)
ridge.fit(x_train,y_train)

print(ridge.coef_)
print(ridge.intercept_)

y_pred3=ridge.predict(x_test)

print(y_pred3-y_test)
print('MAE:',mean_absolute_error(y_pred3,y_test ))
print('MSE:',mean_squared_error(y_pred3, y_test))
print('RMSE:',np.sqrt(mean_squared_error(y_pred3, y_test)))
print('R2:',r2_score(y_pred3, y_test))
```

Ridge model is better than LinearRegression and Lasso Model.

Prediction Comparison Sample

```python
Hours=8.3

Score=ridge.coef_*Hours+ridge.intercept_
print('Ridge Score',Score)

Score2=lr.coef_*Hours+lr.intercept_
print('LR Score',Score2)

Score3=lasso.coef_*Hours+lasso.intercept_
print('Lasso Score',Score3)
```
