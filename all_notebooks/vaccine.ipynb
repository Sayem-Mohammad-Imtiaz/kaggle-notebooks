```python
import matplotlib.pyplot as plt
from matplotlib import style

style.use('fivethirtyeight')

font = {'family': 'serif',
        'weight': 'bold',
        'size': 20}
plt.rc('font', **font)

from datetime import date
import matplotlib.dates as mdates
import datetime as dt
```

```python
df = pd.read_csv('/kaggle/input/covid-vaccination-india-district-wise-data/vaccine_doses_statewise.csv')
```

```python
df.head()
```

```python
df.isnull().sum().sum()
```

```python
df = df.set_index(['State'])
df = df.T.reset_index()
```

```python
df.head()
```

```python
df[['index', 'Punjab']].set_index(['index']).plot(figsize = (15, 8))
plt.axhline(10 ** 6, color = 'maroon', linestyle = ':')
plt.xticks(rotation = 90)
plt.show()
```

```python
df['Punjab']
```

```python
df['Punjab'].shift(1).fillna(0).head()
```

```python
df['Punjab'].head()
```

```python
df['dailyPunjab'] = df['Punjab'] - df['Punjab'].shift(1).fillna(0)
```

```python
df['dailyPunjab']
```

```python
df[['index', 'dailyPunjab']].set_index(['index']).plot(figsize = (15, 8))
plt.xticks(rotation = 90)
plt.show()
```

```python
roll = df['dailyPunjab'].rolling(10, min_periods = 10).mean()
roll[9]
```

```python
roll[9:].plot(figsize = (15, 8))
```

```python
df.iloc[:, 1:] - df.shift().fillna(0).iloc[:, 1:]
```

```python
days = pd.to_datetime(df['index'], dayfirst = True)
days
```

```python
df2 = pd.concat([days, roll], 1)[9:].set_index(['index'])
df2.plot(figsize = (15, 8), linestyle = '--')
plt.vlines(['2021-05-01 00:00:00', '2021-06-21 00:00:00'], ymin = 0.03, ymax = 120000)
#plt.axvline('2021-06-21 00:00:00')
plt.axvspan('2021-05-01 00:00:00', '2021-06-21 00:00:00', color = 'maroon', alpha = 0.3, ymin = 0.04, ymax = 110000)

plt.legend(["Punjab's vaccination drive", "stringent policy of center", "May 1 until June 21, 2021"])
```

```python
df3 = df.iloc[:, 1:37] - df.iloc[:, 1:37].shift(1).fillna(0.0)
```

```python
(pd.concat([days, df3], 1).set_index(['index']) / 10 ** 2).plot(legend = False, figsize = (15, 8), color = '#d9d2d0')
plt.plot(days.values, df['dailyPunjab'].values / 10 ** 2, color = '#0b032b')
#plt.ylim([0, 1750])
```

```python
df4 = df3.rolling(10, min_periods = 10).mean()
temp = pd.concat([days[9:], df4[9:]], 1).set_index(['index'])

string = """The dire performance of 
        vaccination drive
        in Punjab is portrayed 
        from its rolling mean"""

plt.figure(figsize = (15, 8))

plt.plot(days[9:].values, temp.values, color = '#d9d2d0')
plt.plot(days[9:].values, roll[9:].values, color = '#0b032b')

plt.annotate(string, xy = (mdates.date2num(dt.datetime(2021, 3, 1)), 7057.4),
                     xytext = (mdates.date2num(dt.datetime(2021, 1, 30)), 400000),
                     arrowprops = dict(arrowstyle = 'fancy',
                                       color = 'black'),
                     bbox = dict(boxstyle = 'round',
                                 color = 'gray',
                                 alpha = 0.5))

plt.show()
```
