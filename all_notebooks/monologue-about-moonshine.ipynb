```python
mean_cons_per_year = alco.groupby('year', as_index=False).agg({'vodka':'median', 'wine':'median', 'beer':'median', 'champagne':'median', 'brandy':'median'})
```

```python
plt.rcParams['figure.figsize'] = [14, 10]

plt.xlabel('Years')
plt.ylabel('Alcohol consumption in litres per year')

plt.plot(mean_cons_per_year.year, mean_cons_per_year.wine, marker='.', color='red', label='wine')
plt.plot(mean_cons_per_year.year, mean_cons_per_year.vodka, marker='.', color='blue', label='vodka')
plt.plot(mean_cons_per_year.year, mean_cons_per_year.beer, marker='.', color='brown', label='beer')
plt.plot(mean_cons_per_year.year, mean_cons_per_year.champagne, marker='.', color='yellow', label='champagne')
plt.plot(mean_cons_per_year.year, mean_cons_per_year.brandy, marker='.', color='violet', label = 'brandy')
plt.legend()
```

Из этого графика можно сразу сделать некоторые выводы: 
1. Пик потребления пива приходится на период с 2006 по 2014.
2. Падение спроса на водку связано с ростом пивной промешленности в России.
2. На шампанское и коньяк спрос менялся незначительно на протяжении всего рассматриваемого периода. Это в первую очередь связано с ценами на данные виды алкоголя.
3. С 1998 года потребление пива, водки и вина начало расти.
5. Потребление алкоголя значительно уменьшается к 2016 году.

Some conclusions can be drawn from this graph at once:
1. The peak of beer consumption falls on the period from 2006 to 2014.
2. The fall in demand for vodka is due to the growth of the beer industry in Russia.
2. Demand for champagne and brandy varied insignificantly throughout the whole considered period. This is primarily due to the prices for these types of alcohol.
3. Since 1998, the consumption of beer, vodka and wine began to grow.
5. Alcohol consumption will decrease significantly by 2016.


Российская Федерация пережила насколько достаточно мощных кризисов за рассматриваемый период. Давайте их учтём при рассмотрении данных.

The Russian Federation has experienced quite a powerful crisis over the period under review. Let's take them into account when looking at the data.

```python
mean_cons_per_year['crisis_year']='not crisis'
mean_cons_per_year.loc[mean_cons_per_year['year']==1998, 'crisis_year'] = 'crisis'
mean_cons_per_year.loc[(mean_cons_per_year['year']>=2008)&((mean_cons_per_year['year']<=2010)), 'crisis_year'] = 'crisis'
mean_cons_per_year.loc[(mean_cons_per_year['year']>=2014)&((mean_cons_per_year['year']<=2015)), 'crisis_year'] = 'crisis'
```

I really wanted to take into account the income of the population, but the ruble exchange rate changed significantly, as did the prices for food, so, for now, I will only take into account the crisis time.

```python
sns.catplot(x="crisis_year", y="vodka", data=mean_cons_per_year)
sns.catplot(x="crisis_year", y="beer", data=mean_cons_per_year)
sns.catplot(x="crisis_year", y="wine", data=mean_cons_per_year)
sns.catplot(x="crisis_year", y="champagne", data=mean_cons_per_year)
sns.catplot(x="crisis_year", y="brandy", data=mean_cons_per_year)

```

При таком сравнении кажется, что во время экономического кризиса потребление любого алкоголя снижается, однако в данных, которые не касасются кризисного времени для России, мы наблюдаем очень большой разброс данных.

 It seems that during an economic crisis, the consumption of any alcohol decreases, but in the data that do not relate to the crisis time for Russia, we observe a very large scatter of data.
 
Now we need to average the values of consumption of various alcoholic products.

```python
mean_cons_in_crisis = mean_cons_per_year.groupby('crisis_year', as_index=False).agg({'vodka':'median', 'wine':'median', 'beer':'median', 'champagne':'median', 'brandy':'median'})
```

```python
mean_cons_in_crisis.head()
```

```python
plt.rcParams['figure.figsize'] = [10, 10]

plt.xlabel('crisis year')
plt.ylabel('Alcohol consumption in litres per capita')

plt.plot(mean_cons_in_crisis.crisis_year, mean_cons_in_crisis.wine, marker='.', color='red', label='wine')
plt.plot(mean_cons_in_crisis.crisis_year, mean_cons_in_crisis.vodka, marker='.', color='blue', label='vodka')
plt.plot(mean_cons_in_crisis.crisis_year, mean_cons_in_crisis.beer, marker='.', color='brown', label='beer')
plt.plot(mean_cons_in_crisis.crisis_year, mean_cons_in_crisis.champagne, marker='.', color='yellow', label='champagne')
plt.plot(mean_cons_in_crisis.crisis_year, mean_cons_in_crisis.brandy, marker='.', color='violet', label = 'brandy')
plt.legend()
```

**Здесь можно сделать два вывода:**
1. Во время кризиса медианное потребление алкоголя возрастает.
2. Во время кризиса медианное потребление водки падает. Очевидно, что в реальности это не так. В нашем случае падение связано с общим падением потребления водки.

**Two conclusions can be drawn here:**
1. During a crisis, median alcohol consumption increases.
2. During the crisis, the median consumption of vodka falls. This is clearly not the case in reality. In our case, the drop is associated with a general drop in vodka consumption.

Выше я отмечала общее падение медианного потребления алкоголя. У этого есть причина. **На протяжении 12 лет (с 2009) регулирование рынка алкогольной продукции ужесточается.** Растут акцизы, и всё больше крепкого алкоголя уходит в *"серую" зону*, которую, конечно, никто не учитывает. 

Люди пьют, люди пьют очень много, они не станут резко пить меньше.**Но повышение цен на алкоголь и снижение доходов вынуждает их покупать алкоголь в местах, где его продают нелегально.** Контрафактный алкоголь, особенно водка, убивает людей в России, но борьба с ним бессмысленна, пока доходы населения будут продолжать падать, а цены будут расти.

Above, I noted a general drop in median alcohol consumption. There is a reason for this. **For 12 years (since 2009), the regulation of the alcoholic beverages market has been tightening.** Excise taxes are growing, and more and more strong alcohol goes into the *"gray"* zone, which, of course, no one takes into account.

People drink, people drink a lot, they won't drink too much. **But rising alcohol prices and lower incomes are forcing them to buy alcohol in places where it is sold illegally. Counterfeit alcohol, especially vodka, is killing people in Russia, but fighting it is pointless as long as incomes continue to fall and prices rise.
