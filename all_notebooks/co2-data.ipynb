```python
populations = merged_df["Population"]
country_names = merged_df["Country"]
co2_values = merged_df["Co2"]
```

```python
populations
```

```python
co2_per_capita = []
for i in range(len(populations)):
    try:
        co2_per_capita.append(co2_values[i] / populations[i])
    except TypeError:
        co2_per_capita.append(np.nan)
```

```python
merged_df["Co2 per Capita"] = co2_per_capita
```

```python
mean_co2 = merged_df["Co2 per Capita"].mean()
```

```python
fig, ax = plt.subplots(figsize=(16,9))
colors = []
sizes = []
countries_diagramm = []
ax.set_yscale('log')


for x, y, label in zip(populations, co2_values, country_names):
    if type(x) == float and type(y) == float:
        if y > 500_000_000 or (y / x) > 100 or y / x < 0.1:

            # this method is called for each point
            ax.annotate(label, # this is the text
                         (x,y), # this is the point to label
                         textcoords="offset points", # how to position the text
                         xytext=(random.randint(-5, 5),random.randint(0, 15)), # distance from text to points (x,y)
                         ha='center')
            countries_diagramm.append(label)
        

colors, sizes ,x , y , _ = zip(*list(map(lambda x: ["red", 4] + list(x) if x[2] in countries_diagramm else ["blue", 0.8] + list(x), filter(lambda x: x[2] in countries_diagramm or random.random() > 0.8, zip(populations, co2_values, country_names)))))
ax.scatter(x, y, s=sizes, c=colors)

regression_xs = list([i for i in range(1, int(1.4 * 10**9), 10**6)])
def get_regr_y(x):
    return x * mean_co2
regression_ys = list(map(get_regr_y ,regression_xs))
ax.plot(regression_xs, regression_ys, label="best fit line")
ax.legend()
```
