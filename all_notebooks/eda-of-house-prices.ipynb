```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from scipy import stats
```

```python
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv("../input/house-prices/House_price.csv")
```

```python
data.head()
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

```python
data1 = data.Address
```

```python
data1 = data1.str.split(expand= True)
```

```python
#extracting the data regarding state from the given Address column
a = len(data1)
for i in range(a):
    for j in range(9):
        if data1.iloc[i,j] != None:
            if len(data1.iloc[i,j])>2:
                data1.iloc[i,j] = None
```

```python
data1['State'] = data1[data1.columns[1:]].apply(
    lambda x: ','.join(x.dropna().astype(str)),
    axis=1)
```

```python
data2 = data1.State
```

```python
data = pd.concat([data, data2], axis=1)
```

```python
data['Price'].plot.hist(bins = 20)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Price')
plt.show()
```

```python
x = data['Avg. Area Income']
y = data['Price']
plt.scatter(x,y)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color = 'red')
plt.ylabel('House Price')
plt.xlabel('Area Income')
plt.title("Price vs Area Income")
plt.show()
```

```python
x = data['House Age']
y = data['Price']
plt.scatter(x,y)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color = 'red')
plt.ylabel('House Price')
plt.xlabel('Age of House')
plt.title("Price vs Age of House")
plt.show()
```

```python
x = data['Number of Rooms']
y = data['Price']
plt.scatter(x,y)
m, b = np.polyfit(x, y, 1)
plt.ylabel('House Price')
plt.xlabel('No of Rooms')
plt.plot(x, m*x + b, color = 'red')
plt.title("Price vs No of Rooms")
plt.show()
```

```python
x = data['Number of Bedrooms']
y = data['Price']
plt.scatter(x,y)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color = 'red')
plt.ylabel('House Price')
plt.xlabel('No of Bedrooms')
plt.title("Price vs No of Bedrooms")
plt.show()
```

```python
x = data['Area Population']
y = data['Price']
plt.scatter(x,y)
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x + b, color = 'red')
plt.ylabel('House Price')
plt.xlabel('Area Population')
plt.title("Price vs Area Population")
plt.show()
```

```python
x = data['State']
y = data['Price']
fig = plt.figure(figsize=(35, 10))
ax = fig.add_subplot(121)
ax.scatter(x, y)
plt.title("Price vs State")
plt.show()
```

```python
#from the Price vs State graph is seems quite clear that distribution across each state is fairly normal and state 
# doesn't seem to be having any effect on the final price of the house 
# Hence, one can drop the Address part of the data and can proceed without it.
data = data.drop(labels = 'Address', axis = 1)
```

```python
#we export the data into a csv file
data.to_csv("House_price_6.csv", index = False)
```
