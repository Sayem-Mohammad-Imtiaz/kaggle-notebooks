```python
# Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Importing data set

customers_df = pd.read_csv('../input/mall-customers/Mall_Customers.csv')
```

```python
customers_df.head(3)
```

```python
customers_df.info()
```

```python
x = customers_df[['Annual Income (k$)' , 'Spending Score (1-100)']]

x.shape
```

```python
# Visualize the data

plt.scatter(x['Annual Income (k$)'] , x['Spending Score (1-100)'] , color = 'grey')
```

```python
# Finding the optimal no.of clusters

from sklearn.cluster import KMeans

wcss = []
for i in range(1,30):
    kmeans = KMeans(n_clusters= i , init = 'k-means++' , random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
# Visualize the No.of Clusters VS WCSS
```

```python
plt.plot(range(1,30) , wcss)
plt.title('Optimal No.of clusters VS WCSS')
plt.xlabel('No.of clusters')
plt.ylabel('WCSS')
```

```python
# Fitting the model

kmeans = KMeans(n_clusters= 5 , init='k-means++' , random_state=42)
y_predict = kmeans.fit_predict(x)

y_predict
```

```python
y_predict.shape
```

```python
# Testing the data set

kmeans.predict(np.asarray([[91,90]]))
```

```python
y1_predcit = kmeans.predict(np.asarray([[12,67]]))

y1_predcit[0]
```

```python
# Visualize the cluster

plt.scatter(x[y_predict == 0]['Annual Income (k$)'] , x[y_predict == 0]['Spending Score (1-100)'] , s=100 , c = 'red' ,label = 'cluster-1')
plt.scatter(x[y_predict == 1]['Annual Income (k$)'] , x[y_predict == 1]['Spending Score (1-100)'] , s=100 , c = 'blue' , label = 'cluster-2')
plt.scatter(x[y_predict == 2]['Annual Income (k$)'] , x[y_predict == 2]['Spending Score (1-100)'] , s=100 , c = 'green' , label = 'cluster-3')
plt.scatter(x[y_predict == 3]['Annual Income (k$)'] , x[y_predict == 3]['Spending Score (1-100)'] , s = 100 , c = 'yellow' , label = 'cluster-4')
plt.scatter(x[y_predict == 4]['Annual Income (k$)'] , x[y_predict == 4]['Spending Score (1-100)'] , s = 100 , c = 'pink' , label = 'cluster-5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'cyan', label = 'Centroids')
plt.title('Cluster of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
