# **Scale data using StandardScaler**

```python
X = StandardScaler().fit_transform(data)
X
```

# **Use k-Means clustering**

```python
kmeans = KMeans(n_clusters = 10)
model = kmeans.fit(X)
model
```

```python
centers = model.cluster_centers_
centers
```

# **Function for DataFrame and fucntion for plotting**

```python
# create DataFrame with a column for cluster number
def pd_centers(featuresUsed, centers):
    col_names = list(featuresUsed)
    col_names.append("prediction")
    
#     Zip with a column called "prediction" (index)
    Z = [np.append(A, index) for index, A in enumerate(centers)]
#     print(Z)
    
#     Convert to pandas DataFrame for plotting
    P = pd.DataFrame(Z, columns = col_names)
    P["prediction"] = P["prediction"].astype(int)
    return P
```

```python
# function that creates Paralell Plots
def parallel_plot(data):
    my_colours = list(islice(cycle(['b', 'r', 'g', 'y', 'k']), None,len(data)))
    plt.figure(figsize=(15,8)).gca().axes.set_ylim([-5,+5])
    parallel_coordinates(data, "prediction", color = my_colours, marker = "o")
```

```python
P = pd_centers(data.columns, centers)
P
```

```python
# Plotting for strong wine
parallel_plot(P[P["Alcohol"] > 0.5])
```

```python
# Plotting for strong color instensity
parallel_plot(P[P["Color_Intensity"] > 0.5])
```

```python
# Plotting for strong ash level
parallel_plot(P[P["Ash"] > 0.5])
```
