```python
data=pd.read_csv('../input/passenger-list-for-the-estonia-ferry-disaster/estonia-passenger-list.csv')
```

```python
Age=data['Age'].values
cage=[]
for i in Age:
    if i<=16:
        cage.append('Child')
    elif i<=60:
        cage.append('Adult')
    else:
        cage.append('Old Person')
```

```python
data['Age_Category']=cage
```

```python
data.head()
```

```python
# WE are gonna find out the survival ratio for different Age Groups 
```

```python
dic={}
for i in ['Child','Adult','Old Person']:
    dic[i]=len(data[(data['Age_Category']==i) & (data['Survived']==1)].values)/len((data[data['Age_Category']==i] ).values)
```

# Logistic Regression

```python
data.head()
```

```python
df=data
```

```python
df.drop(columns=['Firstname','Lastname','PassengerId'],inplace=True,axis=1)
```

```python
df.head()
```

```python
def makenum(text):
    dic={}
    count=0
    for i in set(list(df[text].values)):
        dic[i]=count
        count+=1
    df[text].replace(dic,inplace=True)
```

```python
df.head()
```

```python
makenum('Country')
```

```python
makenum('Sex')
makenum('Category')
makenum('Age_Category')
```

```python
df.head()
```

```python
X=df.drop(columns=['Survived'],axis=1).values
y=df['Survived'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split?
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.40, random_state=64)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
y_hat=log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print('The score of the model is:',accuracy_score(y_hat,y_test))
```
