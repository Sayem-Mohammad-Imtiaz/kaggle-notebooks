```python
data = pd.read_csv('../input/biomechanical-features-of-orthopedic-patients/column_2C_weka.csv')

data.head()
```

```python
#creating x and y data

# %%
data['class'] = [1 if each =="Normal" else 0 for each in data['class']]
y = data['class'].values
x_data = data.drop(['class'], axis=1)

x_data
```

```python
# %% normalization
# We scale the x data to the range between 0 and 1

x=(x_data-np.min(x_data))/(np.max(x_data)-np.min(x_data)).values

#  (x- min(x))/(max(x)-min(x))
```

```python
# %% train test splits:
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
# knn fitting
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 15)
knn.fit(x_train,y_train)
prediction=knn.predict(x_test)
```

```python
# %% accuracy:
print('{} nn score: {}'.format(15, knn.score(x_test, y_test)))
```

```python
# %% find k value
score_list=[]
#   append depolamaya yarar.
for each in range(1,25):
    knn2=KNeighborsClassifier(n_neighbors=each)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))

plt.plot(range(1,25),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```
