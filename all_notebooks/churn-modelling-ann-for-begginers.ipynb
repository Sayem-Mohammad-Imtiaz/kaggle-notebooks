# Preprocessing 

**Looking at the data we can see that there object types of data in the column Gender and Geography** so we have to convet them into numericals 

```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
data['Gender']=le.fit_transform(data['Gender'])
```

Lets check the data once again 

```python
data.head()
```

We can see that the object type is converted into numeric. The value 0 is assigned to female and 1 to male 

Since Male and female are related we use label encoder but for geography we will be using one hot encoding as they are not related to each other 

```python
data=pd.get_dummies(data,columns=['Geography'])
```

Look at the data again and check if the object types are converted to numeric 

```python
data.head()
```

# Selecting independent and dependent variables 

Row Numbers do not effect to target variable so drop it.
* Do customer id and surname effect the target variable ? Think about it and then proceed further.



* Those do not effect the target variable so lets drop it 

```python
x=data[['CreditScore','Gender','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary',
       'Geography_France','Geography_Germany','Geography_Spain']]
```

# NOTE

The value for france is [1,0,0] ,for Germany [0,1,0] and Spain [0,0,1]

```python
y=data[['Exited']]
```

# Splitting train and test data and feature scaling

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=7)
```

```python
print(xtrain)
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
xtrain=ss.fit_transform(xtrain)
xtest=ss.fit_transform(xtest)
```

# Model building 

```python
from keras.models import Sequential
from keras.layers import Dense 
```

```python
model =Sequential()
```

```python
xtrain.shape
```

12 columns are the input dimensions for the model 

```python
model.add(Dense(10,activation='relu',input_dim=12)) #this the input layer and the first hidden layer of our network
```

```python
model.add(Dense(12,activation='relu'))
model.add(Dense(1,activation='sigmoid'))  #output layer
```

Sigmoid activation is due to binary classification 

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

Adam is the best optimizer , for binary classification the loss function used is binary crossentropy

# Fitting the model

```python
model.fit(xtrain,ytrain,batch_size=32,epochs=100)
```

```python
# testing the model with test data 

ypred =model.predict(xtest)
ypred=(ypred>0.5)
```

```python
#Metrics for the above model 

from sklearn.metrics import  classification_report,confusion_matrix
print(classification_report(ypred,ytest))
```

```python
confusion_matrix(ypred,ytest)
```

```python
from sklearn.metrics import accuracy_score as acs
print(acs(ypred,ytest))
print(acs(ypred1,ytest))
```
