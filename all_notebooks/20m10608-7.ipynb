# Create a global map for COVID Vaccination(total vaccinations)
Total vaccinations represent the country proceed vaccinations physically and politically.
(some media said that Japan is slow to vaccinate.)
According to the figure, China is most advanced in vaccination.It means that China get(make or import) vaccines a lot and proceed very quickly.

```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv', header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
#print(df.index)
#print(df.columns)
range_max = max(df['total_vaccinations'])
fig = px.choropleth(df, locations='country', locationmode='country names', color='total_vaccinations', hover_name='country', animation_frame='date', color_continuous_scale='rainbow', range_color=(0.0, range_max))
fig.update_layout(title_text="total_vaccinations", title_x=0.5)
fig.show()
#plotly.offline.plot(fig, filename='vaccinations.html')
```
