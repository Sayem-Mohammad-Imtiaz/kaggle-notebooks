#### > No null values found

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('dark_background')
```

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.pointplot(x='State/UTs',y='Total Cases',data=covid)
```

#### > Maharashtra has highest number of total cases

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.pointplot(x='State/UTs',y='Active',data=covid,color='yellow')
```

#### > Maharashtra has highest number of active cases

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.pointplot(x='State/UTs',y='Discharged',data=covid,color='green')
```

#### > Maharashtra has highest number of discharged patients

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.pointplot(x='State/UTs',y='Deaths',data=covid,color='red')
```

#### > Maharashtra has highest number of deaths

```python
df=covid.iloc[:,0:5]
df.head()
```

```python
plt.figure(figsize=(12,5))
sns.pointplot(data=df,color='purple')
```

#### > Number of Total Cases and Discharged are more than that for Active and Deaths

```python
plt.figure(figsize=(12,5))
sns.lineplot(data=df,palette='rainbow')
```
