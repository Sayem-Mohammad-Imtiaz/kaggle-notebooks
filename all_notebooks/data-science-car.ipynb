**Kendi istediğimiz araba markası ve modeline göre filtreleme**

Ford marka , Mustang model araba isteğimize göre bulunan arabalardan ilk sıradaki aracın özelliklerini aşağıda yazdırdık.

```python
#kendi isteğimize özel filtreleme yapıp ilk veriyi yazıdırdık
data_ozel = data[np.logical_and(data['car'] == 'Ford' , data['model'] == 'Mustang')]
for index, value in data_ozel[0:1].iterrows():
    print(index, ' : ', value)
```

**Fonksiyon ile isteklerimize göre Araba listeletme **

Aşağıda araba markası default olarak BMW verilmiş olan fonksiyonda model ve engV seçimimize göre arabaları listeletiyoruz.
İstediğimiz doğrultusunda  olan  arabalar  listenedi.


```python
engV = 5.0 # global
#user defined fonction
def Bmw(model, engV, car = 'BMW'):
    """BMW marka arabaların istenilen model ve motor hacmine göre listelenmesi"""
    data_ozel = data[(data['car'] == car) & ( data['model'] == model) & (data['engV'] == engV)]
    return data_ozel
data_ozel =Bmw('M5',engV)
data_ozel
```

**List Comprehension ile yeni sütun ekleme **

Kilometresi ortlama kilometreden daha yüksek olanlara 'high mileage', daha düşük olanlara da 'low mileage' yazdırdık.

```python
ort = sum(data['mileage']) / len(data['mileage'])
print("ortalama kilometre: ", ort)
data['mileage_level'] = ["high mileage" if i > ort else "low mileage" for i in data['mileage']]
data.loc[:10]
```

Yılı 2010'dan düşük olanlara 'eski model', 2010 - 2016 arasındakilere 'yeni model', diğer durumlar içinde 'son model' yazdırdık.

```python
data['year_model_level'] = ["eski model" if i<2010 else "yeni model" if(2016 > i) else "son model" for i in data['year']]
data.head(20)
```
