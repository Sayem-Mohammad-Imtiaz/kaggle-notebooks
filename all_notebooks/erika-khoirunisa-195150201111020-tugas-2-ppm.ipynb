```python
# melakukan Encoding nominal data
df["species"].replace({
    'Adelie': 0, 
    'Gentoo': 1, 
    'Chinstrap': 2
}, inplace=True)

df["island"].replace({
    'Biscoe': 0, 
    'Dream': 1, 
    'Torgersen': 2
}, inplace=True)

df["sex"].replace({
    'MALE': 0, 
    'FEMALE': 1
}, inplace=True)
```

```python
df.info()
```

```python
def rentang_data(df_input):
    list_feature = df.select_dtypes(float)

    for i in list_feature:
        min = df_input[i].min()
        max = df_input[i].max()
        print("Rentang data {fitur} adalah {nilai}".format(fitur=i, nilai=(max-min)))
```

```python
rentang_data(df)
```

```python
def minmax(df_input):
    df_input = df_input.copy()
    list_feature = df_input.select_dtypes(float)

    for i in list_feature:
        min = df_input[i].min()
        max = df_input[i].max()
        df_input[i] = (df_input[i] - min) / (max - min)
    return df_input
```

```python
df_minmax = minmax(df)
df_minmax
```

```python
rentang_data(df_minmax)
```
