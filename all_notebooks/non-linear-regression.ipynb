# feature engineering

```python
new_df = df2.copy()
```

```python
new_df['brand']=new_df['name'].str.split(' ').str.get(0)
new_df.drop(['name'],axis=1,inplace=True)
```

```python
new_df = new_df[new_df['selling_price']<2500000]
new_df['selling_price'] = np.log(new_df['selling_price'])
new_df = new_df[new_df['km_driven']<300000]
#new_df = new_df[new_df['max_power']<300]
new_df['max_power'] = np.log(new_df['max_power'])
new_df['age'] = np.log(new_df['age'])
new_df = new_df[~new_df['fuel'].isin(['CNG','LPG'])]
new_df['mileage'] = new_df['mileage']/70 #to maximize range
```

```python
new_df['seats'] = str(new_df['seats'])
new_df['owner'] = new_df['owner'].astype(object)
```
