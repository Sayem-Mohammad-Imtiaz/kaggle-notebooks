```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot
```

```python
df=pd.read_csv('../input/tipping/tips.csv')
df.head(5)
```

```python
df.info()
```

```python
df.shape
```

```python
# Describe the attributes
df.describe()
```

```python
# there is no missing values
df.isnull().sum()
```

```python
#Number of the smoker by categorized
df['smoker'].value_counts()
```

```python
df['sex'].unique()
sex_smoker=pd.crosstab(df['sex'],df['smoker'])
sex_smoker
```

```python
# Make a filter with male and the tip is more than $5
df[(df['sex']=='Male') & (df['tip']>5)]
```

```python
df[(df['size']>4) & (df['day']=='Sat')]
```

```python
df.groupby('day').mean().sort_values('size',ascending=True)
```

```python
pd.pivot_table(df,index=['day','time'],values='tip')
```

```python
import matplotlib.pyplot as plt
sns.countplot(data=df,y='time',palette='hls')
plt.title('time by categorized')
plt.figure(figsize=(10,5))
plt.show()
```

```python
sns.countplot(data=df,y='smoker',hue='time')
plt.title('who smokes')
plt.figure(figsize=(10,5))
plt.show()
```

```python
sns.relplot('day','tip',data=df,kind='line',ci=None)
```

```python
sns.catplot('sex','size',data=df,kind='box')
```

```python
sns.relplot('tip','size',data=df, kind='line')
```

```python
sns.regplot('tip','size',data=df,color='pink')
```

```python
sns.jointplot('tip','size',data=df,kind='reg',color='red')
```

```python
fig,ax=plt.subplots(figsize=(10,5))
sns.countplot(df['sex'],hue=df['day'],ax=ax)
plt.xlabel('tip')
plt.ylabel('sex')
plt.xticks(rotation=50)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(10,5)
df['time'].value_counts().sort_values(ascending=True).plot.bar(color='yellow')
plt.xlabel('time')
plt.ylabel('count')
plt.show()
```

```python
plt.rcParams['figure.figsize']=(10,5)
df['day'].value_counts().sort_values(ascending=True).plot.pie(y='day',autopct="%0.1f%%")
plt.title('day of the week ')
plt.axis('off')
plt.show()
```

```python
nums=['tip','total_bill']
for i in nums:
    sns.jointplot(x=df[i],y=df['size'], kind='kde',color='red')
    plt.xlabel('i')
    plt.ylabel('count')
    plt.xticks()
    plt.show()
```
