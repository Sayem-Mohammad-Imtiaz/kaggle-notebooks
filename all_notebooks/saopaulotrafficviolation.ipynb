**Import files of violtions** 
infracoes_m_2017.CSV = manual fines applied by enforcement agent
infracoes_e_2017.csv = automatic fines applied by enforcement equipment

```python
#import files to data frame and merge it

dfm2016 = pd.read_csv("../input/sao-paulo-traffic-violations/infracoes_m_2016.csv", sep = ",")
dfe2016 = pd.read_csv("../input/sao-paulo-traffic-violations/infracoes_e_2016.csv", sep = ",")

dfm2017 = pd.read_csv("../input/sao-paulo-traffic-violations/infracoes_m_2017.CSV", sep = ",")
dfe2017 = pd.read_csv("../input/sao-paulo-traffic-violations/infracoes_e_2017.csv", sep = ",")

dfm2018 = pd.read_csv("../input/sao-paulo-traffic-violations/infracoes_m_2018.csv", sep = ",")
dfe2018 = pd.read_csv("../input/sao-paulo-traffic-violations/infracoes_e_2018.csv", sep = ",")



df = dfe2017.append(dfm2017)
df = df.append(dfe2018)
df = df.append(dfm2018)
df = df.append(dfm2016)
df = df.append(dfe2016)

df.sample(5)
```
