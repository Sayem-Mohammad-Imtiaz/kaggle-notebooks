```python
import pandas as pd
import math
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%reload_ext autoreload
%autoreload 2
```

# Imporing dataset

```python
df_crime=pd.read_csv("../input/up-crime-data-year-2019/MR data - Compiled Data Set.csv")
df_crime.head(5) # printing first five rows of the csv file
```

# Preprocessing Dataset

```python
df_crime.shape   # shape(dimension) of data
```

```python
df_crime.isnull().values.any()  #checking if there is any null value in the data
```

```python
df_crime.fillna(0,inplace=True, downcast='infer') #to keep the changes in df specify "inplace=True" and to make 0 integer use downcast
```

```python
df_crime.drop(["Title","Text"],axis=1,inplace=True)   # droping two columns named Title and text
```

```python
df_crime.shape   #after deleting two columns again checking shape(dimension) of data 
```

```python
df_crime=df_crime.rename(columns={'''Murder Happened with reason:\n1:Property/Land Disputes,\n2:Family Dispute\n3:Petty Quarrels,\n4:Money Disputes, \n5:Personal Vendetta, \n6:Love Affairs, \n7:Casteism\n8: Unknown/other''':"murder_reasons",
                         
                         
                         '''Crime Against Women ( Combined ):\n1. Murder with Rape \n2. Dowry Deaths(Sec. 3048)\n3. Suicide(sec 305/306)\n4. Kidnapping(All)\n5. Acid Attack(Sec. 326A IPC)\n6. Cruelty by Husband/in-laws((Sec.498 A IPC)\n7. Rape only(Sec. 376 or 511 IPC)\n8. Assault on Women with Intent to Outrage her Modesty (Sec. 354 IPC)\n9. Cyber Crimes against Women\n10. Protection of Children from Sexual Offences Act''':"ca_women",
                         "Kidnnaping:\nNumber of child victims":"child(kidnapping)",
                         "Number of male(adult) ":"male(kidnapping)",
                         
                         "Number of female(adult)":"female(kidnapping)"})  #changing some columns name for convenience

                         


df_crime.head()
```

```python
df_crime["murder_reasons"].unique()  #just checking what are the murder reasons
```

```python
df_crime[11]=0
df_crime[12]=0
df_crime[13]=0
for i in range(len(df_crime)):                      #spliting columns 8
    if ',' in str(df_crime.iloc[i,8]):
        cl=str(df_crime.iloc[i,8]).split(",")
        df_crime.iloc[i,11]=int(cl[0])
        df_crime.iloc[i,12]=int(cl[1])

        if len(cl)>2:
            df_crime.iloc[i,13]=int(cl[2])


```

```python
df_crime.head()
```

# Drawing plots

```python
#Pie chart graph for the total crimes for each city
df_crime.groupby(['City ']).count()
cities=['Ghaziabad','Kanpur','Lucknow']
total_crimes=np.array([167, 97,202])
plt.pie(total_crimes, labels= cities,autopct='%1.1f%%',startangle=45, radius=1)
plt.show()
```

Observations:
1. Among these three cities lucknow has highest percentage of crime rate while kanpur has the least percentage of crime rate.

```python
#Pie plot for total case reported for each murder reasons 

df_crime.groupby(['murder_reasons']).count()
total_crimes=np.array([14,30,40,19,10,16,16,2])
murder_reasons=['Money Disputes', 'Love Affairs', 'Unknown reasons',
       'Petty Quarrels', 'Property Disputes', 'Family Dispute',
       'Personal Vendetta', 'Casteism']

plt.pie(total_crimes,autopct='%1.1f%%',startangle=45,radius=1.5)

plt.legend(labels= murder_reasons,prop={'size': 15},bbox_to_anchor=(1.2, 0, 0.5, 1) )
plt.show()
```

Observations:
1. The graph shows percentage of total crime for different reasons
2. percentage of total murder is higher for some unknown reasons means though case is
   reported but actual reasons for murder is not specified while the second big factor for murder is love affairs 
3. murders dues to casteism has a very percentage 

```python
#Bar plot for total case reported for each murder reasons(it gives exact total number of cases for different murder reasons )

df_crime.groupby(['murder_reasons']).count()
total_crimes=np.array([14,30,40,19,10,16,16,2])
murder_reasons=['Money Disputes', 'Love Affairs', 'Unknown reasons',
       'Petty Quarrels', 'Property Disputes', 'Family Dispute',
       'Personal Vendetta', 'Casteism']



plt.bar(murder_reasons,total_crimes,color="y")
plt.xticks(murder_reasons, rotation=90)

plt.show()
```

Observation:
1. Unlike the pie chart this graph is giving exact number of murders for each murder reasons 



```python
#Graph of city v/s crime typ"
df_crime['kidnapping']= df_crime['child(kidnapping)']+ df_crime["male(kidnapping)"]+df_crime['female(kidnapping)']
df_crime['murder']= df_crime['Number of child victims'] + df_crime['Number of male victims(adult)'] + df_crime['Number of female victims(adult)']
df_crime['crime against women']=df_crime['Number of childs invoved']+df_crime['Number of adult victims']
arr=df_crime.groupby(['City ']).sum().to_numpy()
kid=arr[:,11]
murder=arr[:,12]
women=arr[:,13]

index=['Ghaziabad', 'Kanpur', 'Lucknow']
df = pd.DataFrame({'Kidnapping': kid, 'Murder': murder, 'crime against women': women}, index=index)
ax=df.plot.bar(rot=0,figsize=(22,10))
```

Observations:

1.Crime against women in lucknow higher as compared to other two cities    
2.Maximum type of crime committed in each city is crime against women while murder is at the second     
3.Kidnapping cases are very low as compared to murder cases and crime against women    
4.Kidnapping cases in Ghaziabad is comaparatively high
