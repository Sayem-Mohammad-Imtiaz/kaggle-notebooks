```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
import math
```

```python
df = pd.read_csv('../input/students-data-for-mlr/amsPredictionSheet1-201009-150447.csv')
df.head()
```

```python
df.describe()
```

```python
corr = df.corr()
corr.style.background_gradient()
```

```python
endog = df['ESE']
exog = sm.add_constant(df[['MSE','Attendance','HRS']])
print(exog)
```

```python
X = exog.to_numpy()
Y = endog.to_numpy()
```

```python
xt = np.transpose(X)
print(xt)
```

```python
multi = np.matmul(xt,X)
print(multi)
```

```python
inv = np.linalg.inv(multi)
multi2 = np.matmul(inv,xt)
print(multi2)
```

```python
beta = np.matmul(multi2,Y)
print(beta)
```

```python
model = sm.OLS(endog,exog)
results = model.fit()
print(results.summary())
```
