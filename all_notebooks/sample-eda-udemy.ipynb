# UDEMY COURSES DATASET

#### This dataset contains all course data for all subjects.


```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
courses = pd.read_csv('../input/all-major-fields-courses-in-udemy-44k-courses/Udemy_Courses.csv')
```

# cleaning the data.

```python
courses.info()
```

```python
courses.loc[courses.Summary.isnull()]
```

```python
#Replacing the Summary columns NaN values with respect to thre Title.

courses['Summary'].fillna(courses.loc[courses.Summary.isnull()]['Title'],inplace = True)
```

```python
#all set with the nan values

courses.info()
```

# EDA of the datasets

```python
#Which subjects has the maximum number of courses?

plt.figure(figsize=(6,5))
sns.set_style('darkgrid')
ax = sns.countplot(x= courses['Category'])
plt.title('Counts of the Courses')
plt.xlabel('Names of the Courses')
plt.ylabel('counts')

for p in ax.patches:
        ax.annotate('{:}'.format(p.get_height()), (p.get_x()+0.2, p.get_height()+100))
```

### In this we found that the maximum number of courses are of Business and then the Technology. The Finance filed courses are very low


```python
#Which course has the highest rating?

courses.loc[courses.Rating == courses.Rating.max()]
```

```python
#Which 5 courses has the highest rating?

courses.sort_values(by = 'Rating', ascending=False).head()
```

```python
#Which courses has most enrollment?

courses.sort_values(by = 'Enrollment', ascending=False).head(10)
```

```python
stars = courses.groupby(courses['Stars']).count().sort_values(by='Stars',ascending=False)[['Title']].head()
stars.rename(columns = {'Title':'Counts'},inplace = True)
stars
```

```python
#Counts of all the top stars Courses.

sns.barplot(x = stars.index,y = stars.Counts)
plt.show()
```
