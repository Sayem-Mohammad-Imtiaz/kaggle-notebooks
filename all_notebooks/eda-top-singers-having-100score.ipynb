```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
singer = pd.read_csv("../input/best-quality-singer/Project-3 - Merged Data.csv")
```

```python
singer
```

```python
singer["Artist(s)"].value_counts()
```

```python
top = singer.sort_values("Score", ascending = False)  
```

```python
top.dropna(inplace=True)
```

Now we Check how many are having Score of 100. which is 22 entries. so we only took those

```python
top.value_counts("Score")
```

```python
top_1 = top.iloc[1:22]
```

```python
top_1
```

```python
yearly = top_1.sort_values("Year", ascending = True)
```

```python
yearly
```

# TOP - 22 singers from year to year having 100 Score

```python
plt.figure(figsize=(20,14))
sns.pointplot(data=yearly, x='Artist(s)', y='Year', color='red');
plt.tick_params(axis='x', which='major', labelsize=15, rotation=90)
plt.tick_params(axis='y', which='major', labelsize=15)
```
