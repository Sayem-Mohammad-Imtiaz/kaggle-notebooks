There are no exceptional enteries in Cost of product

```python
High_payment = data[data['Cost_of_the_Product']> 214]
```

```python
ax = sns.countplot(x="Reached_on_Time",data=High_payment, palette="icefire",hue="Reached_on_Time")
plt.ylabel("No. of customers")

count4=High_payment['Reached_on_Time'].value_counts()
print(count4)
High_payment['Reached_on_Time'].value_counts(normalize=True)*100
```

57% customer having payment received there delivery on time while 43% cusotmer does not received the delivery on time.

We had calculated the average above that 60% customer from total data got delivery on time Hence we can make a conclution that making payment are getting delay in delivery of product in time
