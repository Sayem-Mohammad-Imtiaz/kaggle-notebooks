# **Read data**

```python
import pandas as pd
gm_data = pd.read_csv('/kaggle/input/tft-match-data/TFT_GrandMaster_MatchData.csv')
gm_data.head()
```

# **Champion and Trait Class**

```python
class Champion:
    def __init__(self,name,stars,items):
        self.name = name
        self.stars = stars
        self.items = items
    
class Trait:
    def __init__(self,name,rank):
        self.name = name
        self.rank = rank
    
```

# **Champion transform functions**

```python
def champion_transform(data_string):
    if len(data_string) < 3: # len == 2 is Empty col
        return []
    champion_list = []
    chmampions = data_string[1:-1].replace('{','').split('},') # [1:-1] removes {} on edges
    
    for i in range(len(chmampions)):
        x=chmampions[i].replace('}','').split('\'')         
        name = x[1]
        stars = int(x[-1][2:])
        items = x[4][2:-2]
        if len(items)>3:
            items = items.replace(']','').replace('[','').replace(' ','').split(',')
            items = list(map(int,items))
        else:
            items = []
        chmp = Champion(name,stars,items)
        champion_list.append(chmp)
    return champion_list   
```

```python
for x in champion_transform(gm_data['champion'][0]):
    print(x.name)
```

```python
gm_data['champ_list'] = gm_data['champion'].apply(champion_transform)
gm_data['champ_list']
```

# **Trait transform function**

```python
def trait_transform(data_string):
    if len(data_string) < 3: # len == 2 is Empty col
            return []
    trait_list = []
    for row in data_string[1:-1].split(','): # [1:-1] removes {} on edges
        x = row.split(':')
        name = x[0].replace('\'','').replace(' ','')
        rank = int(x[1])
        trait = Trait(name,rank)
        trait_list.append(trait)
    return trait_list
```

```python
for x in trait_transform(gm_data['combination'][0]):
    print(x.name,x.rank)
```

```python
gm_data['traits'] = gm_data['combination'].apply(trait_transform)
gm_data['traits']
```

# **Using json library**

```python
import json
json.__version__
```

```python

def champion_transform_json(data_string):
    if len(data_string) < 3: # len == 2 is Empty col
        return []
    champion_list = []    
    data_string = data_string.replace('\'',"\"") #Without changing ' ' -> " ", json throws errors
    data_string = json.loads(data_string)
    for name,values in data_string.items():        
        chmp = Champion(name,values['star'],values['items'])
        champion_list.append(chmp)
    return champion_list        
```

```python
for x in champion_transform_json(gm_data['champion'][0]):
    print(x.name)
```

```python
def trait_transform_json(data_string):
    if len(data_string) < 3: # len == 2 is Empty col
            return []
    trait_list = []
    data_string = data_string.replace('\'',"\"") #Without changing ' ' -> " ", json throws errors
    data_string = json.loads(data_string)    
    for name,value in data_string.items():
        trait = Trait(name,value)
        trait_list.append(trait)
    return trait_list
```

```python
for x in trait_transform_json(gm_data['combination'][0]):
    print(x.name,x.rank)
```

```python
gm_data['champ_list_json'] = gm_data['champion'].apply(champion_transform_json)
gm_data['traits_json'] = gm_data['combination'].apply(trait_transform_json)
```
