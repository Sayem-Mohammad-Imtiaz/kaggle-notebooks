Создать новый столбец, в котором разместить разницу между
максимумом и минимумом.

```python
# data.loc[:, ['budget', 'gross', 'runtime', 'score', 'votes', 'year']]
data['max-min row diff'] = np.ptp(data.loc[:, ['budget', 'gross', # разница между максимумом и минимумом по строкам
                                               'runtime', 'score', 'votes', 'year']].values, axis = 1)

data['max budget diff'] = data.budget.max() - data.budget # разница между максимумом бюджета и текущим значением бюджета

data['min budget diff'] = data.budget - data.budget.min() # разница между текущим значением бюджета и его минимумом

data['diff min max diff'] = data['max budget diff'] - data['min budget diff'] # разница между значениями двух предыдущих колонок

data.head()
```

Разбить данные на несколько блоков (не менее 3, приблизительно равных

Сгруппировать данные (GroupBy) по некоторому признаку и сохранить
результаты в новые таблицы. Для каждой новой таблицы провести
сортировку по сложному ключу, состоящему из нескольких признаков.
Для каждой таблицы своя сортировка.

```python
dfs = {key: data for key, data in data.groupby('genre')}
dfs.keys()
action, musical, crime, family = dfs['Action'], dfs['Musical'], dfs['Crime'], dfs['Family']
```

```python
action.sort_values(by = ['gross', 'budget'], ascending=False) # отсортировать экшн-фильмы по убыванию, 
                                                              # признаки: доход от фильма и бюджет
```

```python
musical.sort_values(by = ['score', 'votes'], ascending = False) # отсортировать мюзиклы по убыванию 
                                                                # по признакам пользовательского рейтинга
                                                                # и количеству проголосовавших пользователей
```

```python
crime.sort_values(by = ['score', 'gross'], ascending = False) # отсортировать "криминал" по убыванию 
                                                              # по признакам пользовательского рейтинга
                                                              # и выручки от фильма
```

```python
family[(family.rating == 'G') | (family.rating == 'PG')].sort_values(by = ['rating', 'score'], ascending = [True, False]) 
# отсортировать семейные фильмы по возрастанию по признаку рейтинга и по убыванию пользовательского рейтинга
```
