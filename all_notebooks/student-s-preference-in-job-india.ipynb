<font size=6 color='blue'>**Insights Regarding the Opinions Of Youngsters Among India**</font>

```python
#Importing Necessary Libraries.

import pandas as pd
import numpy as nm
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set_style('darkgrid')
```

```python
#Loading the DataSet.

employment= pd.read_csv('../input/job-survey-india/JobSurveyIndia.csv')
employment.head()
```

```python
#Basic Information About Dataset.

employment.info()
```

<font size=6>**Cleaning Of Data**</font>

```python
#Dropping the Unnecessary column.

employment.drop(columns='Unnamed: 0',inplace=True)
```

```python
#Replacing Values for Non-Ambiguity Entries.

employment['Employment Background'] = employment['Employment Background'].apply(lambda x:'Art' if x=='Art ' else x )
employment['Degree'] =employment['Degree'].apply(lambda x:'Yes' if x=='Yes ' else x)
employment['Public Dealing'] =employment['Public Dealing'].apply(lambda x:'No' if x=='No ' else x)
```

<font size=6>**Insights**</font>

```python
#percentage of preference of distinct Sector

sec1= employment.loc[employment['Employment Sector']=='Private Sector']
sec2= employment.loc[employment['Employment Sector']=='Government Sector']
sec3= employment.loc[employment['Employment Sector']=='Entrepreneurship']

private= 100*(len(sec1)/len(employment))
govt= 100*(len(sec2)/len(employment))
entre= 100*(len(sec3)/len(employment))

print(private,'% of Students Prefer to be in Private Sector.')
print(govt,'% of Students Prefer to be in Government Sector.')
print(entre,'% of Students Prefer to be a Entrepreuner.')
```

```python
#percentage of preference in Distinct Background

sec1= employment.loc[employment['Employment Background']=='Technical']
sec2= employment.loc[employment['Employment Background']=='Non-Technical']
sec3= employment.loc[employment['Employment Background']=='Art']

Tech= round(100*(len(sec1)/len(employment)),2)
Non_Tech= round(100*(len(sec2)/len(employment)),2)
Art= round(100*(len(sec3)/len(employment)),2)

print(Tech,'% of Students Prefer to go into Technical Field.')
print(Non_Tech,'% of Students Prefer to go into Non-Technical Field.')
print(Art,'% of Students Prefer to go into Art Field.')
```
