# Training Letter-26 


```python
# Importing the libraries
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
```

# Preparing and studying the dataset

```python
data = pd.read_csv('/kaggle/input/letterrecognition-using-svm/letter-recognition.csv')
data.head()
```

## Demonstrate the class distribution

```python
X = data.iloc[:, 1:].values
y = data["letter"].values
X = MinMaxScaler().fit_transform(X)



unique, count = np.unique(y, return_counts=True)
plt.figure(figsize=(17, 2))
plt.bar(unique, count, color='gray', data=data.columns)
plt.title('Class Frequency')
plt.xlabel('Class')
plt.ylabel('Frequency')
plt.show()
```

### Illustrate the Black-and-white rectangular pixel displays some of the 26 capital letters in the English alphabet. 

```python
T = 12
plt.figure(figsize=(17, 17))
for i in range(T):
    img = np.asarray(X[i].reshape((4, 4)))
    plt.subplot(6, 6, i + 1)
    plt.imshow(img, cmap='gist_gray')
```

## Splitting the dataset for validation purpose

```python
rando_state = np.random.randint(20)
x_train, x_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=0.3,
                                                    random_state=rando_state)
```

# Defining the classification model

```python
model = GradientBoostingClassifier(learning_rate=0.1,
                                   subsample=1,
                                   max_features="sqrt",
                                   criterion="mse",
                                   n_estimators=150,
                                   max_depth=3,
                                   random_state=rando_state,
                                   verbose=True)
```

## Training the Gradient Boosting

I trained 150 decision regressor trees as our base learner with three as a depth of each tree.

```python
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
score = model.score(x_test, y_test)
score
```

# Check the loss curve of each additive step

Here we have in train loss for each boosting iteration on the in-bag sample.

```python
plt.plot(model.train_score_)
```
