# Objectives

1. To fond if race/ethnicity affects the total score of student?
2. To find if parents level of education is deciding factor for students score.
3. to find of test preperation courses matter?

```python
df = pd.read_csv('../input/students-performance/StudentsPerformance.csv')
df.head(5)
```

Lets look at the data We have information about students gender , race/ethnicity, parental level of education , lunch , test preperation course and his/her scores in maths , reading and writing.

```python
df.info()
```

```python
df.columns
```

lets rename the column names 


```python
df.columns = ['gender', 'race', 'parent_edu', 'lunch', 'test_prep', 'math_s', 'reading_s', 'writing_s']
df.columns
```

```python
df['gender'].value_counts()
```

lets create the additional column for total score

```python
df['total'] = df.math_s + df.reading_s + df.writing_s
df.head(5)
```

# Male Vs Female scores

1.Male vs Female total score

```python

fig = px.histogram(df, x="total", color="gender")
fig.show()
```

Above graph implies that the distrbution of marks is almost same between male and female. 

Male vs Female test scores according to their race

```python

fig = px.box(df, x="race", y="total", color="gender")
fig.update_traces(quartilemethod="exclusive") 
fig.show()
```

**Inference**
* We observe that there is a similar variability of scores in all the races . Male from group B and C have more outliers as compared to other ethno-gender group. 
* The group E seems to be high performing group amongst all other groups. It has median score of 220 and 221 for male and female respectively. 
* Females from the the group A are worst performers with the median of just 183 for female and 191 for male.
* In all groups except E ,males are performing better than females

##hover over the graph to see the median values*


Male vs Female scores according to their parents education

```python

fig = px.box(df, x="parent_edu", y="total", color="gender")
fig.update_traces(quartilemethod="exclusive") 
fig.show()
```

**Inference**
* Female whose parents have masters degree are performing better than any of the ethno-gender category.
* Students whose parents have only high school education have less score in the exams.



```python
fig = px.violin(df, y="total", x="lunch", color="gender", box=True, points="all",
          hover_data=df.columns)
fig.show()
```

**Inference**
* Student who have Standard have high median score as compared to student who have free/reduced lunch

```python
fig = px.violin(df, y="total", x="test_prep", color="gender", box=True, points="all",
          hover_data=df.columns)
fig.show()
```

**Inference**
* Student who have completed their test preperation have high median score as compared to others.
