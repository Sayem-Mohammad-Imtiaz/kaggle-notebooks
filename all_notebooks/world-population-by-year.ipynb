# Prediction of Population by Year

```python
#import library
import numpy as np
import pandas as pd
import seaborn as sns
import math
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV
from sklearn.metrics import confusion_matrix, accuracy_score, mean_squared_error
```

## Data Extraction

```python
population = pd.read_csv('../input/world-population-by-year/WorldPopulation.csv')
df = pd.DataFrame(population)
```

## Machine Learning Model

```python
#split data
X = df.drop('Population', axis = 1)
y = df['Population']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#linear regression
linreg = LinearRegression()
linreg.fit(X_train, y_train)
print(linreg)
```

```python
#Ridge Regression
ridge = RidgeCV()
ridge.fit(X_train, y_train)
print(ridge)
```

```python
#lasso regression
lasso = LassoCV()
lasso.fit(X_train, y_train)
print(lasso)
```

```python
#linear regression prediction
linreg_pred = linreg.predict(X_test)
print(linreg_pred)
```

```python
#ridge regression prediction
ridge_pred = ridge.predict(X_test)
print(ridge_pred)
```

```python
#lasso regression prediction
lasso_pred = lasso.predict(X_test)
print(lasso_pred)
```

```python
#linear regression accuracy
accuracy = linreg.score(X_test, y_test)
print('Linear Regression Score : ', accuracy)
print('Percentage : ', int(accuracy * 100), '%')
```

```python
#ridge regression accuracy
accuracy = ridge.score(X_test, y_test)
print('Ridge Regression Score : ', accuracy)
print('Percentage : ', int(accuracy * 100), '%')
```

```python
#lasso regression accuracy
accuracy = lasso.score(X_test, y_test)
print('Lasso Regression Score : ', accuracy)
print('Percentage : ', int(accuracy * 100), '%')
```

### Linear Regression is the best result of score
