```python
import pandas as pd
```

```python
path = '../input/en-fr-translation-dataset/en-fr.csv'
nrows = 2_000_000
df = pd.read_csv(path , nrows = nrows).astype(str)
```

```python
def get_len(sentence): return len(sentence.split())
filtered = filter(lambda x: get_len(x[1]) <= 1, df[1:].to_numpy())
data_uncleared = pd.DataFrame(filtered).rename(columns={0: 'enlish', 1:'french'})
```
