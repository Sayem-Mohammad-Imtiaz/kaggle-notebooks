```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("../input/netflix-shows/netflix_titles.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
US=df[df.country=='United States']
```

```python
sns.countplot(x='rating',data=US)
```

```python
countryy = df.groupby(['country'])['show_id'].count().reset_index(name='count').sort_values(by='count',ascending=False)
countryy=countryy.head(10)
countryy
```

```python
plt.figure(figsize=(10,10))
# make barplot and sort bars in descending order
sns.barplot(x='country',
            y='count',
            data=countryy,
            order=countryy.sort_values('count',ascending = False).country)
#set axis for count label
ax = plt.gca()
# set labels
plt.xlabel("country", size=15)
plt.ylabel("Count", size=15)
# Iterate through the list of axes' patches
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()),
            fontsize=10, color='black', ha='center', va='bottom')
plt.title("Sort country in Descending Order", size=18)
plt.tight_layout()
plt.show()
```

```python
sns.countplot(x='type',data=US)
```

```python
plt.figure(figsize=(15,2550))
US_stats =  US[US.release_year >= 2015]
g = sns.catplot(data=US_stats, kind="count",x="release_year", hue="type")
```

```python
movie = US[US.type == 'Movie']
listed = movie.groupby(['listed_in'])['show_id'].count().reset_index(name='count').sort_values(by='count',ascending=False)
listed=listed.head()
```

```python
plt.figure(figsize=(20,10))
# make barplot and sort bars in descending order
sns.barplot(x='listed_in',
            y='count',
            data=listed,
            order=listed.sort_values('count',ascending = False).listed_in)
#set axis for count label
ax = plt.gca()
# set labels
plt.xlabel("listed_in", size=15)
plt.ylabel("Count", size=15)
# Iterate through the list of axes' patches
for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()),
            fontsize=10, color='black', ha='center', va='bottom')
plt.title("Sort country with movie category in Descending Order", size=18)
plt.tight_layout()
plt.show()
```
