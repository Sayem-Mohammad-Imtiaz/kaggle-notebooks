```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib as plt
import re
%matplotlib inline
```

```python
#Read in the Data
dataframe = pd.read_csv("../input/mushrooms.csv")
```

```python
#Have a look at the dataframe
dataframe.head()
```

```python
#Train test split
from sklearn.cross_validation import train_test_split
```

```python
#Make dummies for each of the variables
dummies = pd.get_dummies(dataframe, drop_first=True)
dummies.head()
```

```python
#Create the dataframe for X
X = dummies.drop(["class_p"],axis=1)
```

```python
#Create dataframe for Y
y = dummies["class_p"]
```

```python
#Split into train and testing data
X_train, X_test, y_train, y_test = train_test_split(X=, y, test_size=0.4)
```

```python
#import Logistical regression
from sklearn.linear_model import LogisticRegression
```

```python
#Create logistical model, train the model
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
columns_1 = X.columns.T
```

```python
cdf = pd.DataFrame(logmodel.coef_[0],columns_1, columns = ["Coeff"])
```

```python
cdf.sort_values("Coeff")
```

```python
#Have a look at the most important coefficients:
cdf[abs(cdf["Coeff"])>1].sort_values("Coeff")
```
