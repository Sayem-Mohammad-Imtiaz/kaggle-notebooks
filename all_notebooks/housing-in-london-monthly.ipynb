We have 61.10% missing values for 'no_of_crimes' feature whereas <1% of values are missing for 'house_sold'.  We will need to investigate this further to determine the best way of handling this. 

Let's first convert our Date column into a datetime object

```python
df.date # format is YYYY-MM-DD

#convert col into dt object using pandas
df['date'] = pd.to_datetime(df['date'])
df.date.head()
```

```python
from datetime import datetime
import datetime as dt

df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day_no'] = df['date'].dt.weekday
df['day_name'] = df['date'].dt.day_name()

df.head(20)
```

```python
df['code'].unique()
```

Trends of Crimes per Area 

```python
df.groupby('code')['no_of_crimes'].agg(['min', 'max', 'mean', 'count']).plot(kind='bar', figsize=(20,10))
```

Let's view a histogram of the aggregate amounts

```python
df.groupby('code')['no_of_crimes'].agg(['min', 'max', 'mean', 'count']).hist(figsize=(20,10))
```

Let's look at crimes per year

```python
df.groupby('year')['no_of_crimes'].agg(['min', 'max', 'mean', 'count']).plot(kind='bar', figsize=(20,10))
```
