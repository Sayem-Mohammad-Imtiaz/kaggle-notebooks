```python
# importing required modules

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import column_or_1d
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
```

```python
# loading the dataset

data = pd.read_csv('/kaggle/input/diamonds/diamonds.csv')
data.head()
```

```python
for i in ['cut','color','clarity']:
    print(data[i].unique())
```

```python
data.describe()
```

```python
# check missing data

data.isnull().sum()
```

```python
# abstraction of independent and dependent features

X, y = data.drop(['Unnamed: 0', 'price'], axis=1), data['price']
print(X.shape, y.shape)
```

```python
# encoding categorical data

class MyLabelEncoder(LabelEncoder):
    # overriding the fit method due to need of label encoding in particular order
    def fit(self, y):
        y = column_or_1d(y, warn=True)
        self.classes_ = pd.Series(y).unique()
        return self
    
le = MyLabelEncoder().fit(['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'])
X.loc[:, 'cut'] = le.transform(X.loc[:, 'cut'])

le.fit(['J','I','H','G','F','E','D'])
X.loc[:, 'color'] = le.transform(X.loc[:, 'color'])

le.fit(['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF'])
X.loc[:, 'clarity'] = le.transform(X.loc[:, 'clarity'])

X.head()
```

```python
# splitting the data into training and testing sets

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
print(X_train.shape, X_test.shape)
```

```python
# scaling the features

st = StandardScaler()
X_train = st.fit_transform(X_train)
X_test = st.transform(X_test)
```

```python
# applying the regression model

reg = RandomForestRegressor(max_depth=10).fit(X_train, y_train)
print('Training Set Score: {:.5f}'.format(reg.score(X_train, y_train)))
print('Testing Set Score: {:.5f}'.format(reg.score(X_test, y_test)))
```
