```python
from numpy import array
from numpy import mean
from numpy import cov
from numpy.linalg import eig
import pandas as pd
from sklearn.decomposition import PCA
```

```python
A = array([[2.5,2.4],
          [0.5,0.7],
        [2.2,2.9],
         [1.9,2.2],
       [3.1,3.0],
          [2.3,2.7],
         [2,1.6],
          [1,1.1],
           [1.5,1.6],
          [1.1,0.9]])
```

```python
print(A)
```

```python
M = mean(A.T, axis=1)
print(M)
```

```python
C = A - M
print(C)
```

```python
V = cov(C.T)
print(V)
```

```python
values, vectors = eig(V)
print(vectors)
print(values)
```

```python
P = vectors.T.dot(C.T)
print(P.T)
```

```python
pca = PCA(2)
pca.fit_transform(A)
print(pca.components_)
print(pca.explained_variance_)
```

```python
B = pca.transform(A)
print(B)
```
