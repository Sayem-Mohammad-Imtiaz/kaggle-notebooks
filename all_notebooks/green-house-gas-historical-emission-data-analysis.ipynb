```python
df= df.drop('1990',axis=1)
```

```python
df= df.drop('unit',axis=1)
df.head(5)
```

```python

df['1991'] = df['1991'].fillna(value=df['1991'].mean())
df['1992'] = df['1992'].fillna(value=df['1992'].mean())
df['1993'] = df['1993'].fillna(value=df['1993'].mean())
df['1994'] = df['1994'].fillna(value=df['1994'].mean())
df['1995'] = df['1995'].fillna(value=df['1995'].mean())
df['1996'] = df['1996'].fillna(value=df['1996'].mean())
df['1997'] = df['1997'].fillna(value=df['1997'].mean())
df['1998'] = df['1998'].fillna(value=df['1998'].mean())
df['1999'] = df['1999'].fillna(value=df['1999'].mean())
df['2000'] = df['2000'].fillna(value=df['2000'].mean())
df['2001'] = df['2001'].fillna(value=df['2001'].mean())
df['2002'] = df['2002'].fillna(value=df['2002'].mean())
df['2003'] = df['2003'].fillna(value=df['2003'].mean())
df['2004'] = df['2004'].fillna(value=df['2004'].mean())
df['2005'] = df['2005'].fillna(value=df['2005'].mean())
df['2006'] = df['2006'].fillna(value=df['2006'].mean())
df['2007'] = df['2007'].fillna(value=df['2007'].mean())
df['2008'] = df['2008'].fillna(value=df['2008'].mean())
df['2009'] = df['2009'].fillna(value=df['2009'].mean())
df['2010'] = df['2010'].fillna(value=df['2010'].mean())
df['2011'] = df['2011'].fillna(value=df['2011'].mean())
df['2012'] = df['2012'].fillna(value=df['2012'].mean())
df['2013'] = df['2013'].fillna(value=df['2013'].mean())
df['2014'] = df['2014'].fillna(value=df['2014'].mean())
df['2015'] = df['2015'].fillna(value=df['2015'].mean())
df['2016'] = df['2016'].fillna(value=df['2016'].mean())
df['2017'] = df['2017'].fillna(value=df['2017'].mean())
df['2018'] = df['2018'].fillna(value=df['2018'].mean())
```

```python
df.describe()
```

```python
india = df[df['Country/Region']=='India']
india
```

```python
risk_factors = [rf for rf in india.columns if rf not in ['Country/Region']]
risk_factors
```

```python
average_ghg = [] 
for rf in risk_factors:
    average_ghg.append(india[rf].mean())
average_ghg 
```

```python

n_df = pd.DataFrame(list(zip(risk_factors,average_ghg)),columns=['Risk Factor','Avg. ghg']).sort_values(by='Avg. ghg',ascending=False)
```

```python
import seaborn as sns
plt.figure(figsize=(8,8),dpi=300)
sns.barplot(y='Risk Factor',x='Avg. ghg',data=n_df)
plt.title('Green House Gas Historical Emission Data - INDIA')
plt.xlabel('Avg. ghg')
plt.show()
```

```python
usa = df[df['Country/Region']=='United States']
usa 
```

```python
risk_factors = [rf for rf in usa.columns if rf not in ['Country/Region']]
risk_factors
```

```python
average_ghg = [] 
for rf in risk_factors:
    average_ghg.append(usa[rf].mean())
average_ghg
```

```python

us_df = pd.DataFrame(list(zip(risk_factors,average_ghg)),columns=['Risk Factor','Avg. ghg']).sort_values(by='Avg. ghg',ascending=False)
```

```python
import seaborn as sns
plt.figure(figsize=(8,8),dpi=300)
sns.barplot(y='Risk Factor',x='Avg. ghg',data=us_df)
plt.title('Green House Gas Historical Emission Data - USA')
plt.xlabel('Avg. ghg')
plt.show()
```

```python
#Comparison of Average GHG emission in INDIA v/s USA each Year

india_vs_usa = df[(df['Country/Region']=='India') | (df['Country/Region']=='United States')]
india_vs_usa
```

```python
ghg_india = []
ghg_usa = []

for i in ['India','United States']:
    df = india_vs_usa[india_vs_usa['Country/Region']==i]
    
    if i=='India':
        for rf in risk_factors:
            ghg_india.append(df[rf].mean())
    else:
        for rf in risk_factors:
            ghg_usa.append(df[rf].mean())

in_df1 = pd.DataFrame(list(zip(risk_factors,ghg_india)),columns=['Risk Factor','Avg. ghg']).sort_values(by='Avg. ghg',ascending=False)
us_df2 = pd.DataFrame(list(zip(risk_factors,ghg_usa)),columns=['Risk Factor','Avg. ghg']).sort_values(by='Avg. ghg',ascending=False)
```

```python
fig,axes = plt.subplots(1,2,sharey=True,figsize=(12,6),tight_layout=True,dpi=300)
fig.suptitle('Comparision of Average Green House Gas Historical Emission in INDIA v/s USA')
sns.barplot(ax=axes[0],y='Risk Factor',x='Avg. ghg',data=in_df1,palette='viridis')
axes[0].set_title('Avg. ghg in INDIA')
axes[0].set_xlabel('GHG')
sns.barplot(ax=axes[1],y='Risk Factor',x='Avg. ghg',data=us_df2,palette='viridis')
axes[1].set_title('Avg. ghg in USA')
axes[1].set_xlabel('GHG')
axes[1].set_ylabel('')
plt.show()
```
