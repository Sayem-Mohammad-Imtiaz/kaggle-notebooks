```python
import cv2
import pytesseract
```

```python
from PIL import Image, ImageDraw, ImageFont
```

```python
im = cv2.imread('/kaggle/input/handwriting-recognition/train_v2/train//TRAIN_00001.jpg')
config = ('-l eng --oem 1 --psm 3')
text = pytesseract.image_to_string(im, config=config)
text = text.split('\n')
```

```python
text
```

```python
def remove_items(test_list, item): 
    res = [i for i in test_list if i != item] 
    return res
```

```python
text=remove_items(text,'\x0c')
text=remove_items(text,'')
text=remove_items(text,' ')
```

```python
text
```

```python
text_img=""
for i in text:
    text_img=text_img+i+" "
text_im2=""
for i in range(len(text_img)):
    text_im2=text_im2+text_img[i]+' '
    
```

```python
text_im2
```

```python
img = Image.new('RGB', (170, 30), color = 'white')
```

```python
d = ImageDraw.Draw(img)
d.text((10,10), text_im2, fill=(0,0,0))
```

```python
img_width=170
img_height=30
img_color='white'
```

```python
def generate_img(img_width,img_height,img_color,text):
    img_size=(img_width,img_height)
    img = Image.new('RGB', img_size, color = img_color)
    d = ImageDraw.Draw(img)
    d.text((10,10), text, fill=(0,0,0))
    return img
    
    
```

```python
def get_text(img_path):
    im = cv2.imread(img_path)
    config = ('-l eng --oem 1 --psm 3')
    text = pytesseract.image_to_string(im, config=config)
    text = text.split('\n')
    text=remove_items(text,'\x0c')
    text=remove_items(text,'')
    text=remove_items(text,' ')
    text_img=""
    for i in text:
        text_img=text_img+i+" "
    text_im2=""
    for i in range(len(text_img)):
        text_im2=text_im2+text_img[i]+' '
    return text_im2
    
    
    
    
```
