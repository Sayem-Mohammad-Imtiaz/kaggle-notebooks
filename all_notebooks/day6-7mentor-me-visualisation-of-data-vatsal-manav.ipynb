```python
import pandas as pd
df = pd.read_csv('../input/solar-power-generation-data/Plant_1_Weather_Sensor_Data.csv')
df.head()
```

```python
df.nunique()
```

```python
df.info()
```

```python
df['DATE_TIME'] = pd.to_datetime(df['DATE_TIME'],format = '%Y-%m-%d %H:%M')
```

```python
df['DATE'] = pd.to_datetime(df['DATE_TIME'],format = '%Y-%m-%d %H:%M').dt.date
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.plot(df['DATE_TIME'],df['AMBIENT_TEMPERATURE'],label = 'Ambient')
plt.plot(df['DATE_TIME'],df['MODULE_TEMPERATURE'],label = 'Ambient')
plt.grid()
plt.legend()
plt.margins(0.10)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.plot(df['DATE_TIME'],df['AMBIENT_TEMPERATURE']-df['MODULE_TEMPERATURE'],label = 'difference')

plt.grid()
plt.legend()
plt.margins(0.10)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.plot(df['DATE_TIME'],df['AMBIENT_TEMPERATURE'],label = 'Ambient')
plt.plot(df['DATE_TIME'],df['MODULE_TEMPERATURE'].rolling(window=50).mean(),label = 'module')
plt.plot(df['DATE_TIME'],df['AMBIENT_TEMPERATURE']-df['MODULE_TEMPERATURE'].rolling(window=50).mean(),label = 'difference')
plt.grid()
plt.legend()
plt.margins(0.10)
plt.show()
```
