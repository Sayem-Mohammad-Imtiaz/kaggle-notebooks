```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import neighbors
```

```python
dt = pd.read_csv('/kaggle/input/red-wine-quality-cortez-et-al-2009/winequality-red.csv',sep= ',')
```

```python
dt.head(100)
```

```python
predictors = dt.drop(columns=['quality'])
```

```python
target = dt.quality
```

```python
model = neighbors.KNeighborsClassifier(n_neighbors = 10)
```

```python
knn = model.fit(predictors,target)
```
