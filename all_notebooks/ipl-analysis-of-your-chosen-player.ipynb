```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import os

import numpy as np
```

```python
DIR = '../input/ipl2020/'
```

```python
filename = 'all_matches_details.csv'
df = pd.read_csv(DIR + filename)
```

```python
df_batsman_all = df[ df["batsman"] == mybatsman ]
```

```python
len(df_batsman_all)
```

```python
match_data = pd.read_csv(DIR + 'IPLT2020_ALL_MATCHES.csv')
```

# Runs scored per match

```python
runs = df_batsman_all.groupby(['match_no'])['runs_batsman'].sum() \
.reset_index( name = 'runs_batsman') 
```

```python
runs = runs.merge(match_data)
cols = ["match_no","runs_batsman","1st_innings_team","2nd_innings_team"]
runs = runs[cols]
```

# Balls faced per match

```python
balls = df_batsman_all.groupby(['match_no'])['delivery_no'].count() \
.reset_index( name = 'balls_faced') 
```

```python
runs_balls = runs.merge(balls)
```

```python
runs_balls
```

# Strike Rate per match

```python
runs_balls["strike_rate"] = runs_balls["runs_batsman"] / runs_balls["balls_faced"] *100
```

```python
runs_balls
```

# Runs given per match

```python
runs = df_bowler_all.groupby(['match_no'])['runs_total'].sum() \
.reset_index( name = 'runs_total') 
```

```python
runs = runs.merge(match_data)
cols = ["match_no","runs_total","1st_innings_team","2nd_innings_team"]
runs = runs[cols]
```

```python
runs
```

```python
balls = df_bowler_all.groupby(['match_no'])['delivery_no'].count() \
.reset_index( name = 'balls_bowled') 
```

```python
runs_balls = runs.merge(balls)
```

```python
runs_balls["overs_bowled"] = np.floor(runs_balls["balls_bowled"] /6)
```

```python
runs_balls["economy"] =runs_balls["runs_total"] / runs_balls["overs_bowled"]
```
