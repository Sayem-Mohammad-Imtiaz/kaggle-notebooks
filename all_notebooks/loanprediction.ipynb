```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder,OrdinalEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,confusion_matrix

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('/kaggle/input/loan-prediction-based-on-customer-behavior/Training Data.csv')
df.head()
```

```python
plt.figure(figsize=(20,5))
plt.subplot(1,3,1)
sns.boxplot(df['Married/Single'],df['Income'])

plt.subplot(1,3,2)
sns.countplot(df['Car_Ownership'])

h_counts = df['House_Ownership'].value_counts()
plt.subplot(1,3,3)
plt.pie(h_counts.values,labels=h_counts.index,labeldistance=1.1,wedgeprops={'edgecolor':'red','linewidth':2})
plt.show()
```

```python
df['STATE'].value_counts()
```

```python
le = LabelEncoder()
df['Married/Single'] = le.fit_transform(df['Married/Single']) 
```

```python
order_label = {'rented':1,'owned':2,'norent_noown':3}
df['House_Ownership'] = df['House_Ownership'].map(order_label)
```

```python
df['Car_Ownership'] = le.fit_transform(df['Car_Ownership']) 
```

```python
df.info()
```

```python
x = df.iloc[:,[1,2,3,4,5,6,10,11]]
y = df.iloc[:,[-1]]
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=55)
```

```python
print("X_train shape =",x_train.shape)
print("X_test shape =",x_test.shape)
print("Y_train shape =",y_train.shape)
print("Y_test shape =",y_test.shape)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
y_pred_lr = lr.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred_lr))
print(accuracy_score(y_test,y_pred_lr).round(2)*100)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion='entropy',random_state=53)
```

```python
dtc.fit(x_train,y_train)
```

```python
y_pred_dtc = dtc.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred_dtc))
print(accuracy_score(y_test,y_pred_dtc).round(2)*100)
```

# Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfr = RandomForestClassifier(n_estimators=100,criterion='gini',random_state=53)
```

```python
rfr.fit(x_train,y_train)
```

```python
y_pred_rfr = rfr.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred_rfr))
print(accuracy_score(y_test,y_pred_rfr).round(2)*100)
```

# K Nearest Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knc = KNeighborsClassifier(n_neighbors=5)
```

```python
knc.fit(x_train,y_train)
```

```python
y_pred_knc = knc.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred_knc))
print(accuracy_score(y_test,y_pred_knc).round(2)*100)
```

# Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
y_pred_nb = nb.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred_nb))
print(accuracy_score(y_test,y_pred_nb).round(2)*100)
```

## Prediction Accuracy

```python
print("Logistic regression:- ",accuracy_score(y_test,y_pred_lr).round(2)*100)
print("Decision Tree Classifier:- ",accuracy_score(y_test,y_pred_dtc).round(2)*100)
print("Random Forest Classifier:- ",accuracy_score(y_test,y_pred_rfr).round(2)*100)
print("K Nearest Neighbors Classifier:- ",accuracy_score(y_test,y_pred_knc).round(2)*100)
print("Naive Bayes Classifier:- ",accuracy_score(y_test,y_pred_nb).round(2)*100)
```
