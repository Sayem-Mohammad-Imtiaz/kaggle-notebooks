```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/upwork-1/management.csv')
```

```python
df
```

```python
df = df.drop_duplicates(['recno'])
```

```python
df
```

```python
df['rate_in_usd'] = df['hourlyRate'].str.replace('{','').str.replace('}', '').str.split('\'amount\':').str[1]
```

```python
df
```

```python
#Most common job titles in dataset
df['title'].value_counts()[0:20].plot(kind='bar')
```

```python
#Find value counts for rates ($/hr) in listings with 'logo' in title
df[df['title'].str.lower().str.contains('logo')]['rate_in_usd'].value_counts()
```

```python
#Common rates in dataset
df['rate_in_usd'].value_counts()[0:20].plot(kind='bar')
```

```python
#Change type
df['rate_in_usd'] = df['rate_in_usd'].astype(float)
```

```python
#Narrow down by rates $30/hr and less
df1 = df[df['rate_in_usd'] <= 30]
```

```python
#Check for employees that...
#1. have 'law' in their title/desc
#2. charge less than $30/hr
df1[df1['title'].str.lower().str.contains('law')]
```
