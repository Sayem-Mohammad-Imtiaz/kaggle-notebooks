```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df=pd.read_csv('../input/excess-mortality-during-the-covid19-pandemic/excess_mortality.csv')
```

```python
df.isnull().sum()
```

```python
df.head()
```

Let's check whether missing values (2020, 2021 and average deaths) for countries are distributed equally or not. 

```python

df_reduced=df[['location','deaths_2020_all_ages','average_deaths_2015_2019_all_ages','deaths_2021_all_ages']].copy()
```

```python
country=df_reduced.groupby('location')['deaths_2020_all_ages'].sum().index.to_list()
death=df_reduced.groupby('location')['deaths_2020_all_ages'].sum().values
death2=df_reduced.groupby('location')['average_deaths_2015_2019_all_ages'].sum().values
death3=df_reduced.groupby('location')['deaths_2021_all_ages'].sum().values
```

```python
for countries in country:
    print('Missing values for ' + str(countries) + ':' + str(df[df.location==countries]['deaths_2020_all_ages'].isnull().sum()))
```

```python
for countries in country:
    print('Missing values for ' + str(countries) + ':' + str(df[df.location==countries]['average_deaths_2015_2019_all_ages'].isnull().sum()))
```

```python
for countries in country:
    print('Missing values for ' + str(countries) + ':' + str(df[df.location==countries]['deaths_2021_all_ages'].isnull().sum()))
```

```python
df_reduced
```

```python
fig=go.Figure([go.Bar(x=country,y=death,name='Deaths at 2020 for all Ages'),
              go.Bar(x=country,y=death2,name='Average Death for All Ages' ),
              go.Bar(x=country, y=death3,name='Deaths at 2021 for all Ages')])
fig.update_layout(barmode='group',
                 height=500,
                 width=1000)
fig.show()
```

Lets consider countries with more mortality to get rid of the noise that have been created by the data representing countries which have less mortality.

```python
df_reduced2=df_reduced[df_reduced.deaths_2020_all_ages>=50000]
```

```python
px.pie(df_reduced2,names=df_reduced2.location, values="deaths_2020_all_ages")
```

```python
country_1=df_reduced2.groupby('location')['deaths_2020_all_ages'].sum().index.to_list()
death_1=df_reduced2.groupby('location')['deaths_2020_all_ages'].sum().values
death2_1=df_reduced2.groupby('location')['average_deaths_2015_2019_all_ages'].sum().values
death3_1=df_reduced2.groupby('location')['deaths_2021_all_ages'].sum().values

fig=go.Figure([go.Bar(x=country_1,y=death_1,name='Deaths at 2020 for all Ages'),
              go.Bar(x=country_1,y=death2_1,name='Average Death for All Ages' ),
              go.Bar(x=country_1, y=death3_1,name='Deaths at 2021 for all Ages')])
fig.update_layout(barmode='group',
                 height=500,
                 width=1000)
fig.show()
```

Since 2021 is not overed yet and vaccines are available currently, lower mortality is expected.
