### What is the name of the person who paid the highest ADR(Average Daily Rate)? How much their ADR?

```python
df.nlargest(1, 'adr')['adr']
```

### What is the mean of the adr?

```python
np.round(df['adr'].mean(),2)
```

### What is the average number of nights for a stay across the entire dataset? (round to 2 decimal points)

```python
df.columns
df['total_stays'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
round(df['total_stays'].mean(),2)
```

### What are the the names and emails of the people who made 5 "Special Requests" ?

```python
df[df['total_of_special_requests'] == 5][['name', 'email']]
```

### What percentage of hotel stays were classified as "repeat guests"? 

```python
round(100*(df[df['is_repeated_guest']!= 0].shape[0])/(df.shape[0]),2)
```

### What are the top 5 most common last names in the dataset? 

```python
df['name'].str.split().str[1].value_counts().head()
```

### What are the names of the people who had booked the most number of children and babies for their stay?

```python
df['offspring'] = df['children'] + df['babies']
df.sort_values('offspring', ascending=False)[['name','offspring']]
```
