```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly as py
import plotly_express as px
import plotly.graph_objects as go
from matplotlib import pyplot as plt
import folium
from folium import plugins
from plotly.offline import init_notebook_mode, iplot
import os
import base64
init_notebook_mode()

df = pd.read_csv('../input/top-women-chess-players/top_women_chess_players_aug_2020.csv')
```

# Top Chess Players by Federation (Country)

```python
df_fed = pd.DataFrame(df.groupby('Federation').size()).reset_index()
df_fed.rename(columns = {0:'Number of Chess Players'}, inplace=True)
fig = px.choropleth(df_fed, locations="Federation",
                    color="Number of Chess Players", # lifeExp is a column of gapminder
                    hover_name="Federation", # column to add to hover information
                    color_continuous_scale=px.colors.sequential.Plasma,
                    title = 'Top Ranked Female Chess Players by Federation')
fig.show()


df_fed.sort_values(by = 'Number of Chess Players', ascending=False, inplace=True)
fig = px.bar(df_fed[0:20], x = 'Federation', y = 'Number of Chess Players', title = 'Top 20 Federations by Number of Top Ranked Female Chess Players')
fig.show()


df_gm = df[df.Title == 'GM']
df_gm_fed = pd.DataFrame(df_gm.groupby('Federation').size()).reset_index()
df_gm_fed.rename(columns = {0:'Number of Grandmasters'}, inplace=True)
fig = px.choropleth(df_gm_fed, locations="Federation",
                    color="Number of Grandmasters", # lifeExp is a column of gapminder
                    hover_name="Federation", # column to add to hover information
                    color_continuous_scale=px.colors.sequential.Plasma,
                    title = 'Female Grandmasters by Federation')
fig.show()


df_gm_fed.sort_values(by = 'Number of Grandmasters', ascending=False, inplace=True)
fig = px.bar(df_gm_fed[0:20], x = 'Federation', y = 'Number of Grandmasters', title = 'Federations Ranked by Number of Female Chess Grandmasters')
fig.show()
```

# Top Chess Players by Birth Year/Age
Difficult to say if the players are deceased or not, so we'll just use birth year

```python
fig = px.histogram(df, x = 'Year_of_birth', title = 'Age Distribution of Top Female Chess Players')
fig.show()

df_title_avg_yob = df[['Title', 'Year_of_birth']].groupby(['Title']).mean().reset_index()
fig = px.bar(df_title_avg_yob, x = 'Title', y = 'Year_of_birth', title = 'Average Year of Birth by Title')
fig.update_yaxes(tickvals=[1978, 1982, 1986, 1990, 1994], range=(1978, 1994))
fig.show()
```
