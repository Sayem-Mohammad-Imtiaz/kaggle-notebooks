# **Predicting if the person will buy Insurance using Logistic Regression**

**Importing Packages**

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

**Importing Insurance data**

```python
data = pd.read_csv("../input/insurance_data.csv")
```

**Basic Analysis**

```python
data.shape
```

```python
data.head()
```

**Separating Feature and Target matrix**

```python
X = data[['age']]
y = data.bought_insurance
```

**Plotting _Age_ vs _Brought Insurance_**

```python
plt.scatter(X,y)
```

**Splitting the data into _Train data_ and _Test data_**

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1)
```

**Defining the _Logistic Regression_ model**

```python
model = LogisticRegression()
```

**Training _Logistic Regression_ model**

```python
model.fit(X_train, y_train)
```

**Using the model for predictions**

```python
y_pred= model.predict(X_test)
```

**Finding the acuracy of prediction**

```python
model.score(X_test,y_test)
```

Find the results [TP  TN

                  FP  FN ]

```python
confusion_matrix(y_test,y_pred)
```
