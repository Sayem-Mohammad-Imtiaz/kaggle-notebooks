Data contain 5 object, 3 integer, and 2 floats. Data does not cotain any duplicates values. We can start analysis. As previously explained, there are 6 category of crime in the city of Montreal. We going to group the data based on category of crime and 10 city with highest crime.

```python
crime_trend = crime.groupby('category').agg({'count':sum}).sort_values('count',ascending = False).reset_index()
crime_trend.head(6)
# Plot the pie chart
plt.pie(crime_trend['count'],labels=crime_trend['category'], autopct='%1.1f%%', startangle=15)
 
# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
plt.subplots_adjust(wspace=2, hspace=0.8)
plt.title("Crime Category in Montreal")
# Display the graph onto the screen
plt.show()
```

```python
#top crime city in Montreal
topcity_crime = crime.groupby('city').agg({'count':sum}).sort_values('count',ascending = False).reset_index()
top10_city = topcity_crime[:10]
other_city = topcity_crime[10:].sum()
other_city.city='other'
other_city=pd.DataFrame(other_city).transpose()
other_city.head()
crime_city= pd.concat([top10_city, other_city])
crime_city.head(11)
# Plot the pie chart
plt.pie(crime_city['count'],labels=crime_city['city'], autopct='%1.1f%%', startangle=30)
# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
plt.subplots_adjust(wspace=2, hspace=1)
plt.title("top 10 high crime city in Montreal")
# Display the graph onto the screen
plt.show()
```

The Pie chart shown that The most common crime in montreal is Home Invasion followed by Theft in/from a motor vechicle. The least crime is offense resulting in death. the highest crime city is Montreal city by 67.8% followed by Montreal-Nord. We will focus to Montreal city to understand which most highest crime neighbourhood.

```python
#top crime neighbourhood in Montreal
topnb_crime = crime[crime.city== 'MONTREAL'].groupby('neighbourhood').agg({'count':sum}).sort_values('count',ascending = False).reset_index()
top10_nb = topnb_crime[:10]
other_nb = topnb_crime[10:].sum()
other_nb.neighbourhood='other'
other_nb=pd.DataFrame(other_nb).transpose()
crime_nb= pd.concat([top10_nb, other_nb])
crime_nb.head(11)
# Plot the pie chart
plt.pie(crime_nb['count'],labels=crime_nb['neighbourhood'], autopct='%1.1f%%', startangle=30)
# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')
plt.subplots_adjust(wspace=2, hspace=1)
plt.title("top 10 high crime neighbourhood in Montreal City")
# Display the graph onto the screen
plt.show()
```

In Montreal City, the highest crime neighbourhood is in Plateu Mont-royal, followed by Mercier. Let see the trend 3 of Crime in Montreal city: Home Invasion, Theft in / from a motor vehicle, and Mischief

```python
crime_trend = crime[crime.city== 'MONTREAL']
tablecrime = pd.pivot_table(crime_trend, values='count', index=['date'],
                    columns=['category'], aggfunc=np.sum)
tablecrime.head()
```

```python
tablecrime1 = tablecrime[: :30]
plt.figure(figsize=(20,6))
lw = 2
plt.plot(tablecrime1.index, tablecrime1['Home Invasion'],  color='red', lw=lw, label='Home Invasion' )
plt.plot(tablecrime1.index, tablecrime1['Mischief'],  color='yellow', lw=lw, label='Mischief' )
plt.plot(tablecrime1.index, tablecrime1['Theft in / from a motor vehicle'],  color='blue', lw=lw, label='Theft in / from a motor vehicle' )
plt.xlabel('DATE')
plt.ylabel('crime')
plt.xticks(rotation=90, fontsize=8)
plt.title('Crame Rate in Montreal City')
plt.legend(loc="upper right")
plt.grid()
plt.show()
```

The graph shown that the trend of home invasion and theft in/from a motor vehichle slightly declined while mischief crime trend tend to be similiar trend by the years.
