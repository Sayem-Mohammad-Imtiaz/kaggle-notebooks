# **Let's break down into cities as well (Top 5)******

```python
df['City'].value_counts()[:5]
```

```python
#Let's clean the city only well known templates selected.
def new_city(city_name):
    
    if ("karachi" in city_name) or ("KARACHI" in city_name) or ("Khi" in city_name) or ("Karachi" in city_name) or ("East Karachi" in city_name):
        return "KHI"
    
    elif ("lahore" in city_name) or ("LAHORE" in city_name) or ("Lhr" in city_name):
        return "LHR"
    
    elif ("islamabad" in city_name) or ("ISLAMABAD" in city_name) or ("Isl" in city_name):
        return "ISL"
    
    elif ("rawalpindi" in city_name) or ("RAWALPINDI" in city_name) or ("Rwp" in city_name) or ("pindi" in city_name):
        return "RWP"
    
    elif ("faisalabad" in city_name) or ("FAISALABAD" in city_name) or ("Fsd" in city_name):
        return "FSD"
    
    else:
        return city_name
```

```python
df["New City"]=df["City"].apply(new_city)
```

# ORDER'S INSIGHT WRT WEEK, MONTH AND YEARS

```python
df['Order Date & Time'] = pd.to_datetime(df['Order Date & Time'])
```

```python
df['Date'] = pd.to_datetime(df['Order Date & Time']).dt.date
df['Day'] = df['Order Date & Time'].dt.day_name()
df['Month'] = df['Order Date & Time'].dt.month_name()
df['Year'] = df['Order Date & Time'].dt.year
```

```python
df.head(3)
```

```python
df['Year'].value_counts()
```

```python
explode = (0.2, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')
df['Year'].value_counts().plot(kind='pie', title='Total Number Of Orders Per Year:', autopct='%1.1f%%', explode=explode, shadow=True);
```

```python
Completed_Orders = df['Order Status'] == 'Completed'
Orders = df[Completed_Orders]
print('Number of Completed Orders in a Year\n', Orders['Year'].value_counts())


Cancelled_Orders = df['Order Status'] == 'Cancelled'
C_Orders = df[Cancelled_Orders]
print('Number of Cancelled Orders in a Year\n', C_Orders['Year'].value_counts())


Returned_Orders = df['Order Status'] == 'Returned'
R_Orders = df[Returned_Orders]
print('Number of Returned Orders in a Year\n', R_Orders['Year'].value_counts())
```

Now we got a numbers of Completed, Cancelled and Returned Orders wrt Year. Let's do some Plotting.

```python
Orders['Year'].value_counts().plot(kind='bar',title='Number of Completed Orders Per Year');
```

```python
sns.countplot(x='Year', data=C_Orders, palette='Set2');
```

```python
R_Orders['Year'].value_counts().plot(kind='pie',autopct='%1.1f%%',title='%age of Returned Orders Per Year');
```

2020 was the best Year for Guftugu, In which they sold most number of Books, It might be happened because of Covid-19 but so far 2020 is the Winner.
