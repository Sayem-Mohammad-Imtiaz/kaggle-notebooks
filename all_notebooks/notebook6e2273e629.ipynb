```python
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split as tts

from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('../input/student-alcohol-consumption/student-mat.csv')
```

```python
df.head()
```

# Data Visualization

```python
plt.figure(figsize=(14,12))
sns.heatmap(df.corr(),annot = True)
plt.show()
```

# Preprocessing

### Checking For Missing values

```python
df.isnull().sum()
```

### Encoding

```python
df.dtypes
```

```python
nonnumeric_columns = [df.columns[index] for index,dtype in enumerate(df.dtypes) if dtype == 'object']
```

```python
nonnumeric_columns
```

```python
for column in nonnumeric_columns:
    print(f"{column}:     {df[column].unique()}")
```

```python
df['Mjob'] = df['Mjob'].apply(lambda x : "m_"+x)
df['Fjob'] = df['Fjob'].apply(lambda x : "f_"+x)
df['reason'] = df['reason'].apply(lambda x : "r_"+x)
df['guardian'] = df['guardian'].apply(lambda x : "g_"+x)
```

```python
df
```

```python
dummies_df = pd.concat([pd.get_dummies(df['Mjob']),pd.get_dummies(df['Fjob']),pd.get_dummies(df['reason']),pd.get_dummies(df['guardian'])],axis = 1)
```

```python
df = pd.concat([df,dummies_df],axis = 1)
```

```python
df = df.drop(['Mjob','Fjob','reason','guardian'],axis=1)
```

```python
df
```

```python
nonnumeric_columns = [df.columns[index] for index,dtype in enumerate(df.dtypes) if dtype == 'object']
for column in nonnumeric_columns:
    print(f"{column}:  {df[column].unique()}")
```

```python
encoder = LabelEncoder()
for column in nonnumeric_columns:
    df[column] = encoder.fit_transform(df[column])
```

### Scaling

```python
Y = df['G3']
X = df.drop('G3',axis=1)
```

```python
X
```

```python
scaler = StandardScaler()
X = pd.DataFrame(scaler.fit_transform(X),columns=X.columns)
```

```python
X
```

# Training

```python
x_train,x_test,y_train,y_test = tts(X,Y,train_size = 0.7)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

# Performance


```python
import sklearn
```

```python
model.score(x_test,y_test)
```

```python
y_pred = model.predict(x_test)
```

```python
print(f'Explained variance regression score : {sklearn.metrics.explained_variance_score(y_test, y_pred)}')
print(f'maximum residual error : {sklearn.metrics.max_error(y_test, y_pred)}')
print(f'Mean absolute error : {sklearn.metrics.mean_absolute_error(y_test, y_pred)}')
print(f'Mean squared error : {sklearn.metrics.mean_squared_error(y_test, y_pred)}')
print(f'r2 score : {sklearn.metrics.r2_score(y_test, y_pred)}')
```
