```python
from sklearn import ensemble, preprocessing, metrics
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import f1_score
import numpy as np
import pandas as pd
col_names = ['gender', 'age', 'hypertension', 'heart disease', 'ever married', 'work type', 'residence type',
'average glucose level', 'bmi', 'smoking status', 'stroke']
dataset = pd.read_csv('/kaggle/input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv', header=0, names=col_names)
dataset.head()
bmi_median = np.nanmedian(dataset["bmi"])
complete_bmi = np.where(dataset["bmi"].isnull(), bmi_median,
dataset["bmi"])
dataset["bmi"] = complete_bmi
label_encoder = preprocessing.LabelEncoder()
dataset["gender"] = label_encoder.fit_transform(dataset["gender"])
dataset["ever married"] = label_encoder.fit_transform(dataset["ever married"])
dataset["work type"] = label_encoder.fit_transform(dataset["work type"])+1
dataset["residence type"] = label_encoder.fit_transform(dataset["residence type"])
dataset["smoking status"] = label_encoder.fit_transform(dataset["smoking status"])
dataset_X = pd.DataFrame([dataset["gender"], dataset["age"], dataset["heart disease"], dataset["hypertension"], dataset["ever married"], dataset["work type"], dataset["residence type"], dataset["average glucose level"], dataset["bmi"], dataset["smoking status"]]).T
dataset_y = dataset["stroke"]
train_X, test_X, train_y, test_y = train_test_split(dataset_X, dataset_y,
test_size = 0.3, random_state=1)
# Random Forest Model
forest = ensemble.RandomForestClassifier(n_estimators = 100)
forest_fit = forest.fit(train_X, train_y)
# Prediction
test_y_predicted = forest.predict(test_X)
# Evaluate
accuracy = metrics.accuracy_score(test_y, test_y_predicted)
print("RFC accuracy:", accuracy)
RFC_f1 = f1_score(test_y, test_y_predicted)
print("RFC f1 score:", RFC_f1)
```
