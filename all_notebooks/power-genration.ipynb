# What is the mean value of daily yield?

```python
print("mean value of daily yield in plant 1: {:.2f}".format(plant_1gd["DAILY_YIELD"].mean()))
print("mean value of daily yield plant 2: {:.2f}".format(plant_2gd["DAILY_YIELD"].mean()))
print("mean value of daily yield Both plant 1 and 2: {:.2f}".format((plant_2gd["DAILY_YIELD"].mean()+plant_1gd["DAILY_YIELD"].mean())/2))
```

# What is the total irradiation per day?

```python
print("total irradiation per day plant 1: {:.2f}".format(plant_1wsd["IRRADIATION"].sum()))
print("total irradiation per day plant 2: {:.2f}".format(plant_2wsd["IRRADIATION"].sum()))
print("total irradiation per day Both plant 1 and 2: {:.2f}".format((plant_1wsd["IRRADIATION"].sum()+plant_2wsd["IRRADIATION"].sum())))
```

# What is the max ambient and module temperature?

```python
print("max ambient temperature: {:.2f} and max module temperature: {:.2f} in plant 1".format(plant_1wsd["AMBIENT_TEMPERATURE"].max(),plant_1wsd["MODULE_TEMPERATURE"].max()))
print("max ambient temperature: {:.2f} and max module temperature: {:.2f} in plant 2".format(plant_2wsd["AMBIENT_TEMPERATURE"].max(),plant_2wsd["MODULE_TEMPERATURE"].max()))
```

# What is the maximum/minimum amount of DC/AC Power generated in a time interval/day?

```python
plant_1gd.groupby("DATE_TIME").max()
```

```python
plant1gd=plant_1gd.copy()
plant1gd["DATE"]=plant1gd["DATE_TIME"].apply(lambda x:x.split(" ")[0])
plant1gd["TIME"]=plant1gd["DATE_TIME"].apply(lambda x:x.split(" ")[1])
plant1gd.groupby("DATE")["DC_POWER","AC_POWER"].max()
```

```python
plant1gd.groupby("DATE")["DC_POWER","AC_POWER"].min()
```

```python
pd.concat([plant1gd.groupby("DATE")["DC_POWER","AC_POWER"].max(),plant1gd.groupby("DATE")["DC_POWER","AC_POWER"].min()],axis=1,)
```
