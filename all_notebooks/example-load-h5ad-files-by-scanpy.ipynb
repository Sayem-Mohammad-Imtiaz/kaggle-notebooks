# What is about ?

Data - single RNA-seq genes expressions data. 
Mainly subparts extracted from very huge https://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE139944
which does not fit into 16G memory.

Main part of files stored in h5ad files and contain "adata" files from scanpy anndata. 
Here we give examples how to load these files. 
And show how to do some simple manipulations. 

It is quite simple: 

adata.X - count matrix (might be sparse), where column names - genes names, rows - cells

adata.obs - information on cells (same size as X.shape[0] ) - it is dataframe

adata.var - information on genes (same size as X.shape[1] )  - it is dataframe

---------------------------------

K562,MCF7, A549 - names of well-known cell-lines 

HDACs - drugs of type "Histone deacetalise inhibitors"
