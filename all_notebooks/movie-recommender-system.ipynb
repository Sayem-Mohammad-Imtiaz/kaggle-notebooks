Combine Two Dataset

```python
movie_rating_df = pd.merge(movie_rating_df ,movie_title_df,on = 'item_id')
```

```python
movie_rating_df.drop('timestamp',axis=1,inplace=True)
```

```python
dataset = movie_rating_df.groupby('title')['rating'].describe()
```

```python
dataset = dataset.reset_index()
```

```python
dataset = dataset[['title','count','mean']]
```

Modelling

```python
matrix = movie_rating_df.pivot_table(index='user_id',columns='title',values='rating')
```

```python
matrix
```

```python
x = matrix['Titanic (1997)']
```

```python
x
```

```python
x_correlations = pd.DataFrame(matrix.corrwith(x),columns=['Correlation value'])
```

```python
x_correlations
```

```python
x_correlations.dropna(inplace=True)
```

```python
x_correlations
```

```python
x_correlations = pd.merge(x_correlations, dataset, on = 'title')
```

```python
x_correlations
```

```python
x_correlations = x_correlations.sort_values('Correlation value',ascending=False)
```
