```python
df_250 = pd.read_csv('/kaggle/input/restaurant-business-rankings-2020/Top250.csv',index_col=0)
df_250.head()
```

```python
df_250.isna().sum()
```

```python
df_250 = df_250.drop(['Content','Headquarters'],axis=1) #since over 70% of these data pts are missing
df_250.head()
```

```python
def remove_punctuation(column):  #trying to remove the % sign and dividing by 100 since it is percentage
    no_punc = [x for x in column if x not in "%"]
    no_punc = "".join(no_punc)
    divide_100 = float(no_punc)/100
    return divide_100
```

```python
df_250['new_sales_perc'] = df_250['YOY_Sales'].apply(remove_punctuation)  #applying the function
```

```python
df_250.head()
```

```python
df_250['new_sales'] = df_250['new_sales_perc']*df_250['Sales'] +df_250['Sales']  #adding the increased/decreased new values 
                                                                                 
```

```python
df_250.head()
```

```python
df_250['new_units_perc'] = df_250['YOY_Units'].apply(remove_punctuation) #doing the same with units as well
```

```python
df_250.head()
```

```python
df_250['new_units'] = df_250['new_units_perc']*df_250['Units'] +df_250['Units']
```

```python
df_250['new_units'] = df_250['new_units'].astype('int') #since they are units ,convert to integer
```

```python
df_250.head()
```

```python
def gain_loss(column):  #calculating gain or loss using the '-' sign
    for x in str(column):
        if(x.startswith('-')):
            return 0  #0 means we have incurred loss
        else:
            return 1  #1 means we have gains
```

```python
df_250['gl_sales'] = df_250['new_sales_perc'].apply(gain_loss)
```

```python
df_250['gl_units'] = df_250['new_units_perc'].apply(gain_loss)
```

```python
df_250.head()
```

```python
# time to drop the useless columns to keep our dataframe clean
df = df_250.drop(['YOY_Sales','YOY_Units','new_sales_perc','new_units_perc'],axis=1)
df.head()
```

```python
df['Segment_Category'].value_counts()  #getting a count of each unique food type
```

```python
# dividing the plots into 2 spaces and putting the plots next to each other
plt.subplot(1,2,1)
plt.title("Gain/Loss in Sales")
sns.countplot(df['gl_sales'])

plt.subplot(1,2,2)
plt.title("Gain/Loss in Units")
sns.countplot(df['gl_units'])

plt.tight_layout()  #to give space between the two plots  -- feel free to run the code before and after
```
