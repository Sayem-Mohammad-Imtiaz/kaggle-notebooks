```python
# graph
import networkx as nx
from pyvis.network import Network
```

```python
# create graph from edges
state= df["state"].value_counts()

G = nx.from_pandas_edgelist(df, 'state', 'GENUS', create_using=nx.DiGraph())
# and plot
nx.draw(G, with_labels=True, node_size=1000, alpha=0.5, arrows=True)
plt.title('States ' + str(state) + ' Genus')
plt.show()
```

```python
# create graph from edges
GENUS = df['GENUS'].value_counts()[:10]

G = nx.from_pandas_edgelist(df, 'state', 'GENUS', create_using=nx.DiGraph())

# and plot
nx.draw(G, with_labels=True, node_size=1000, alpha=0.5, arrows=True)
plt.title('Fish Genus ' + str(GENUS))
plt.show()
```

#Bidirectional

```python
# create graph from edges
species = df["species"].value_counts()[:10]

G = nx.from_pandas_edgelist(df, 'state', 'GENUS', create_using=nx.DiGraph())

# and plot
nx.draw_circular(G, with_labels=True, node_size=1000, alpha=0.5, arrows=True)
plt.title('Bidirectional view of species ' + str(species))
plt.show()
```

#Interactive Plot

```python
net = Network(notebook=True, width='800px', height='500px')
net.directed = True
net.from_nx(G)
net.show('example.html')
```

#Global Statistics

```python
# frequency table
f_counts = df.species.value_counts().rename_axis('species').reset_index(name='Frequency')

# top 10
f_counts[0:10]
```

```python
# stats
f_counts.Frequency.describe()
```

#Interpretation: A fish has in average 199.69  (mean).

```python
# plot in log scale
plt.hist(np.log10(1+f_counts.Frequency),100)
plt.yscale('log')
plt.grid()
plt.title('Distribution of Endangered Fish Species count')
plt.xlabel('log10(1+counts)')
plt.show()
```
