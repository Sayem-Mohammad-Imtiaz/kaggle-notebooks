* **Merge data**

```python
import pandas as pd
import glob

path = r'../input/brazil-fire-points/BRAZIL_FIRE_POINTS' 
all_files = glob.glob(path + "/*.csv") #array with all csv files

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
list(frame.columns.values)
```

```python
frame.shape
```

**Create YEAR aaaa and MONTH mm columns**

```python
frame['aaaa'] = frame['datahora'].str.slice(0, 4)
frame['mm'] = frame['datahora'].str.slice(5, 7)

frame
```

```python
frame.to_csv('./firepoints.csv')
```

**CALCULATE the monthly average by Month and State**

```python
listaestados=[]
for linha in frame['estado']:
    if [linha] not in listaestados:
        listaestados.append([linha])
lista = frame.values.tolist()
lista
```

```python
dt=[]
for mes in range(1,13):
    for estado in listaestados:
        for ano in range (2018,2020): #change as your selection
            soma = 0 #number of fire points per state per month per year
        
            for item in lista:            
                if (item[3]==estado[0]) and (int(item[13])==mes and (int(item[12])==ano)):
                    soma = soma + 1
            dt.append([estado[0], ano, mes, soma])

df=pd.DataFrame(dt)
df
```

**Count fire points by Year, Month and State**

```python
df.columns = ['estado', 'ano','mes','count']
df
```
