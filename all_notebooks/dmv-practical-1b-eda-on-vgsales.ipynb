```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("../input/videogamesales/vgsales.csv")
df.head()
```

```python
df.describe()
```

```python
year_freq = pd.crosstab(index=df['Year'], columns='count')
temp = []
for i in year_freq['count']:
  p = 100 * (i/df.shape[0])
  temp.append(p)
year_freq['Percent'] = temp
year_freq.sort_values('count', ascending = False)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(year_freq.index,year_freq['count'])
plt.xlabel('year')
plt.ylabel('frequency')
plt.show()
```

```python
plt.plot(year_freq.index,year_freq['Percent'])
plt.show()
```

```python
global_sales_freq = pd.crosstab(index=df['Global_Sales'], columns='count')
f = plt.figure()
f.set_figwidth(15)
f.set_figheight(5)
ax = f.add_axes([0,0,1,1])
ax.bar(global_sales_freq.index,global_sales_freq['count'])
plt.xlabel('total worldwide sales (in millions)')
plt.ylabel('frequency')
plt.show()
```

```python
import numpy
from numpy import percentile
```

```python
sales_loc = ['NA_Sales','EU_Sales','JP_Sales','Other_Sales','Global_Sales']
indices = ['min','Q1','median','Q3','max','iqr','midrange']
summary_data = pd.DataFrame(columns=sales_loc, index=indices)
for i in sales_loc:
  temp = []
  quartiles = percentile(df[i], [25, 50, 75])
  data_min, data_max = df[i].min(), df[i].max()
  temp.append(data_min)
  temp.extend(quartiles)
  temp.append(data_max)
  temp.append(quartiles[2]-quartiles[0])
  temp.append((data_min + data_max)/2)
  summary_data.loc[:,i] = temp

summary_data
```

```python
for i in sales_loc:
  fig = plt.figure(figsize=(10,5))
  low = summary_data.loc['Q1',i] - (1.5 * summary_data.loc['iqr',i])
  high = summary_data.loc['Q3',i] + (1.5 * summary_data.loc['iqr',i])
  plt.xlim(low, high+0.5)
  plt.xticks(np.arange(low, high+0.5, 0.1))
  plt.ylabel(i)
  plt.boxplot(df[i], vert=False)
  plt.show()
  print("\n")
```
