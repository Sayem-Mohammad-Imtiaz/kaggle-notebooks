```python
#Dropping Adj Close because it's the same as close, thus being redundant, in this data set
newData = data.drop(['Adj Close'], axis = 1)
newData.head()
```

```python
#I don't know what I expected checking for correlation...
correlation = newData.corr()
correlation.head()
```

```python
#Again, don't even know why I bothered, there's a 1:1 correlation between OPEN and HIGH....Because stock prices generally don't stretch to far from their open/close
sns.heatmap(correlation,xticklabels = correlation.columns, yticklabels= correlation.columns, annot=True)
```

```python
#Just from looking at this there's no correlation between a stock price and volume on itself. 
sns.relplot(data =newData, x='Volume',y= 'High')
```

```python
#I wasn't expecting the line to go like this....

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math

#set variables need to be in specific format 
X1 = newData['Volume'].values.reshape(-1,1)
y1 = newData['High'].values.reshape(-1,1)

#create train / test split for validation 
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.3, random_state=0)
        
reg = LinearRegression().fit(X_train1, y_train1)
reg.score(X_train1, y_train1)
reg.coef_
y_hat1 = reg.predict(X_train1)

plt.scatter(X_train1,y_train1)
plt.scatter(X_train1,y_hat1)
plt.xlabel( 'VOLUME',fontsize =20)
plt.ylabel('HIGH', fontsize=20)
plt.show()

y_hat_test1 = reg.predict(X_test1)
plt.scatter(X_test1, y_test1)
plt.scatter(X_test1, y_hat_test1 )
plt.xlabel('VOLUME',fontsize =20, c='red')
plt.ylabel('HIGH', fontsize=20)
plt.show()
```
