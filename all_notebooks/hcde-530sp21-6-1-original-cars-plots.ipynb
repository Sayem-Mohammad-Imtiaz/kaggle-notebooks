```python
new_cars = pd.read_json("../input/cars-json/cars.json")
new_cars
```

```python
import plotly.express as px
fig = px.scatter(x=new_cars['Miles_per_Gallon'], y=new_cars['Horsepower'])
fig.show()
```

```python
fig2 = px.scatter(new_cars, x=new_cars['Horsepower'], y=new_cars['Miles_per_Gallon'] , color="Cylinders", hover_name="Name", hover_data=["Origin", "Year"])
fig2.show()
```
