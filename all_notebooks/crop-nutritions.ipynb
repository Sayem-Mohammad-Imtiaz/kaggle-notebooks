```python
def extractFirstWord(st):
    st = st.strip()
    startInd = 0
    l = len(st)
    endInd = 0
    while endInd < l and st[endInd].isalpha():
        endInd += 1
    
    return st[startInd:endInd]
```

```python
extractFirstWord('Barbados Cherry')
```

```python

df['CropSimpleName'] = df['Crop'].apply(lambda x: extractFirstWord(x))
```
