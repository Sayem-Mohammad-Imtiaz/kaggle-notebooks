```python
import pandas as pd 
import numpy as np 
import pylab as pl 
import matplotlib.pyplot as plt 
from sklearn.tree import DecisionTreeClassifier
```

```python
my_data=pd.read_csv('/kaggle/input/heart-attack-analysis-prediction-dataset/heart.csv')
```

```python
my_data.head()
```

```python
my_data.describe()
```

```python
X = my_data[['age','sex','cp','trtbps','chol','fbs','restecg','thalachh','exng','oldpeak','slp','caa','thall']].values
```

```python
X.shape
```

```python
y = my_data['output']
y.shape
```

```python
plt.scatter(my_data.age, my_data.output)
plt.xlabel('Age')
plt.ylabel('Output')
plt.show()
```

```python
import seaborn as sns
sns.set(rc={'figure.figsize':(7,5)})
plt.hist(my_data.age, bins=100)
plt.xlabel('Age')
plt.ylabel('Number of People')
```

<h2> Using Decision Tree

```python
from sklearn.model_selection import train_test_split
X_trainset, X_testset, y_trainset, y_testset = train_test_split(X, y, test_size = 30, random_state = 5)
```

```python
y_testset.shape
```

```python
y_trainset.shape
```

```python
outputTree = DecisionTreeClassifier(criterion='entropy', max_depth=12)
outputTree.fit(X_trainset, y_trainset)
```

```python
predTree = outputTree.predict(X_testset)
```

```python
from sklearn import metrics
print('Accuracy: ',metrics.accuracy_score(y_testset, predTree))
```

<h4> Accuracy comes out to be 0.9
