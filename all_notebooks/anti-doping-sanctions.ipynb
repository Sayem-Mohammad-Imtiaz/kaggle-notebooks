```python
import holoviews as hv
from holoviews import opts
hv.extension('bokeh')
```

```python
df = pd.read_csv("/kaggle/input/cusersmarildownloadssanctioncsv/sanction.csv", delimiter=';', encoding='ISO-8859-2')
df.head()
```

```python
df.isnull().sum()
```

```python
df["sport"].value_counts()
```

```python
#Code by Kohei-mu https://www.kaggle.com/koheimuramatsu/industrial-accident-causal-analysis/notebook

#Fixed by Des desalegngeb https://www.kaggle.com/desalegngeb
#https://www.kaggle.com/desalegngeb/bokeh-visualization-library-guide-for-beginners

sport_cnt = np.round((df['sport'].value_counts(normalize=True) *100).head()) 
hv.Bars(sport_cnt).opts(title="Doping Violation by Sport", color="red", xlabel="Sport", ylabel="Percentage", yformatter='%d%%')\
                .opts(opts.Bars(width=500, height=300,tools=['hover'],show_grid=True)).opts(xrotation=45)
```

#To avoid clumsy plotting, select a few columns

```python
ax = sns.countplot(x = 'sport',data=df,order=['Track and Field','Cycling', 'Mixed Martial Arts','Weightlifting', 'Swimming'])
for p in ax.patches:
        ax.annotate(format(p.get_height()), (p.get_x()+0.2, p.get_height()))
plt.xticks(rotation=45)
plt.title('Doping Violations by Sport');
```

```python
#Code by Taha07  https://www.kaggle.com/taha07/data-scientists-jobs-analysis-visualization/notebook

color = plt.cm.Reds(np.linspace(0,1,20))
df["sport"].value_counts().sort_values(ascending=False).head(10).plot.pie(y="sanction_announced",colors=color,autopct="%0.1f%%")
plt.title("Doping Violation by Sports")
plt.axis("off")
plt.show()
```

```python
#word cloud
from wordcloud import WordCloud, ImageColorGenerator
text = " ".join(str(each) for each in df.sport)
# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=200,colormap='Set2', background_color="Red").generate(text)
plt.figure(figsize=(10,6))
plt.figure(figsize=(15,10))
# Display the generated image:
plt.imshow(wordcloud, interpolation='Bilinear')
plt.axis("off")
plt.figure(1,figsize=(12, 12))
plt.show()
```
