**Importando las librerias**

```python
import matplotlib.pyplot as plt
import pandas as pd 
```

**Cargando el dataset y dividiendo nuestros valores independientes (x) y nuestros valores dependientes(y)**

```python
dataset = pd.read_csv('../input/salary-data-simple-linear-regression/Salary_Data.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

**Cargando la libreria sklearn con el paquete model_selection para traer la función que nos dividira entre los valores de entrenamiento y test**
**IMPORTANTE: Para agregar más parametros a nuestra división de test y entrenamiento se puede leer la documentación**

A Parte de esto llamamos de la libreria sklearn la función LinearRegression (Esta sirve para regresión lineales simples y multiples para conocer esto más a profundidad se puede leer la documentación); tambien crearemos un regresor con el cual ajustaremos el modelo para x_train y para y_train

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state = 0 )

from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(x_train,y_train)
```

Con la variable y_pred crearemos nuestra predición para la cual haremos uso del segmento de test y haremos uso de matplotlib para graficar nuestros resultados primeramente con los datos de entrenamiento para saber que tan acertivo fue

```python
y_pred = regression.predict(x_test)

plt.scatter(x_train, y_train, color="red")
plt.plot(x_train, regression.predict(x_train), color="blue")
plt.title("Sueldo en relación con los años del conjunto de entrenamiento")
plt.xlabel("Años de experiencia")
plt.ylabel("Salarios")
plt.show()
```

Seguido de esto probaremos nuestro modelo con los conjuntos de prueba, con el cual podemos observar que tenemos unos resultados bastante favorables

```python
plt.scatter(x_test, y_test, color="red")
plt.plot(x_train, regression.predict(x_train), color="blue")
plt.title("Sueldo en relación con los años del conjunto de entrenamiento")
plt.xlabel("Años de experiencia")
plt.ylabel("Salarios")
plt.show()
```
