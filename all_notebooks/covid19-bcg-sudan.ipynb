```python
df = pd.read_csv('../input/hackathon/BCG_country_data.csv')
df.head()
```

```python
sudan = df[(df['country_name']=='Sudan')].reset_index(drop=True)
sudan.head()
```

```python
plt.figure(figsize=(24, 6))
plt.subplot(131)
sns.countplot(sudan['gni_per_capita_usd'],order = sudan['gni_per_capita_usd'].value_counts(dropna=False).index)
plt.xticks(rotation=90)
plt.subplot(132)
sns.countplot(sudan['income_group_type'],order = sudan['income_group_type'].value_counts(dropna=False).index)
plt.subplot(133)
sns.countplot(sudan['population_per_km2'],order = sudan['population_per_km2'].value_counts(dropna=False).index)
```

```python
fig, ax = plt.subplots(1,3, figsize = (20,6), sharex=True)
sns.countplot(x='gni_per_capita_usd',data=sudan, palette="copper", ax=ax[0])
sns.countplot(x='gni_per_capita_usd',hue='stroke_disease_disability', palette="ocean", data=sudan,ax=ax[1])
sns.countplot(x='gni_per_capita_usd',hue='percentage_people_with_diabetes', palette="cubehelix", data=sudan,ax=ax[2])
ax[0].title.set_text('GNI per Capita Count')
ax[1].title.set_text('GNI per Capita Vs Stroke Disease Disability')
ax[2].title.set_text('GNI per Capita Vs Percentage of People with Diabetes')
plt.show()
```

```python
fig = px.bar(sudan, 
             x='gni_per_capita_usd', y='income_group_type', color_discrete_sequence=['crimson'],
             title='Sudans GNI per capita in Covid19 Times', text='population_per_km2')
fig.show()
```
