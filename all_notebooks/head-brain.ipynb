```python
data = pd.read_csv('/kaggle/input/headbrain/headbrain.csv')
data
```

```python
#plt.style.use('ggplot')
plt.hist2d(data['Brain Weight(grams)'], data['Head Size(cm^3)'], color='black')
plt.ylabel('Head Size')
plt.xlabel('Brain Weight')
plt.title('Cross-Refrencing the Frequency Distribution of Size and Weight')
plt.show()
```

```python
#Dividing the datasets
x = data['Brain Weight(grams)'].values
y = data['Head Size(cm^3)'].values
```

```python
#reshaping
x = x.reshape(len(data), 1)
y = y.reshape(len(data), 1)
```

```python
# Spliting the data
xtrain, xtest, ytrain, ytest = train_test_split(x, y, train_size=0.80, random_state=385)
```

```python
# Build Model
model = LinearRegression()
```

```python
# Perform training
model.fit(xtrain, ytrain)
```

```python
#Perform Prediction
ypred = model.predict(xtest)
```

```python
#R2 Score
r2 = r2_score(ytest, ypred)
print(r2)
```

```python
plt.figure(figsize=(15,5))
plt.title("Line of Regression")
plt.subplot(1,2,1)
# Line of Regression (training Samples)
plt.scatter(xtrain, ytrain, color="red", label="Actual Points")
plt.scatter(xtrain, model.predict(xtrain), color="blue", label="Predicted Points")
plt.plot(xtrain, model.predict(xtrain), color="orange", label="Line of Regreession")
plt.ylabel('Head Size')
plt.xlabel('Brain Weight')
plt.legend()
plt.title("Training Samples")
plt.subplot(1,2,2)
# Line of Regression (testing Samples)
plt.scatter(xtest, ytest, color="red", label="Actual Points")
plt.scatter(xtest, model.predict(xtest), color="blue", label="Predicted Points")
plt.plot(xtest, model.predict(xtest), color="orange", label="Line of Regreession")
plt.ylabel('Head Size')
plt.xlabel('Brain Weight')
plt.legend()
plt.title("Testing Samples")
plt.show()
```
