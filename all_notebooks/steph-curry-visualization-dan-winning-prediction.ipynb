# Features
* Season_year : Season, ex) 2020-2021 (str)
* Season_div : Pre season, Regular season, Post Season (str)
* Date : Month and day (str)
* OPP : Opposite (str)
* Result : Won or Loss (str)
* T Score : Score of the Team involved Stephen Curry (int)
* O Score : Score of the other (int)
* MIN : Minutes played in game (float)
* FG : Field goals made - attempted (str)
* FG% : Field goals made, attempted and percentage (float)
* 3PT : Three-point field goals made - attempted (str)
* 3PT% : Three-point field goals made, attempted and percentage (float)
* FT : Free throws made - attempted (str)
* FT% : Free throws made, attempted and percentage (float)
* REB : Rebounds in game (int)
* AST : Assists in game (int)
* BLK : Blocks in game (int)
* STL : Steals in game (int)
* PF : Personal fouls in game (int)
* TO : Turnover in game (int)
* PTS : Points in game (int)

```python
pts = []
for i in df['Season_year'].unique():
    pts.append(df[df['Season_year'] == i]['PTS'].mean())

win = []
for i in df['Season_year'].unique():
    win_count = 0
    for j in df[df['Season_year'] == i]['Result']:
        if j == 'W':
            win_count +=1
    win.append(win_count)

fig,ax = plt.subplots(1,2,figsize=(15,10))
ax[0].plot(range(1, len(pts)+1), pts)
ax[0].set_title('Stephs Points per game over the years', fontsize=14)
ax[0].set_xlabel('Season')
ax[0].set_ylabel('PTS pergame')
ax[1].plot(range(1, len(pts)+1), win)
ax[1].set_title('Stephs winning count over the years', fontsize=14)
ax[1].set_xlabel('Season')
ax[1].set_ylabel('Win per season')
fig.show()
```

```python
pt3m = []
for i in df['Season_year'].unique():
    pt3m.append(df[df['Season_year'] == i]['3PTM'].sum())
    
pt3 = []
for i in df['Season_year'].unique():
    pt3.append(df[df['Season_year'] == i]['3P%'].mean())

fig,ax = plt.subplots(1,2,figsize=(15,10))
ax[0].plot(range(1, len(pt3m)+1), pt3m)
ax[0].set_title('3PT made per season', fontsize=14)
ax[0].set_xlabel('Season')
ax[0].set_ylabel('3PT pergame')
ax[1].plot(range(1, len(pt3)+1), pt3)
ax[1].set_title('3PT% per season', fontsize=14)
ax[1].set_xlabel('Season')
ax[1].set_ylabel('3PT%')
fig.show()
```

```python
opp = []
win_count = []
for i in df['OPP'].unique():   
    opp.append(i)    
    win=0
    for j in df[df['OPP'] == i]['Result']:
        if j == 'W':
            win +=1
        else:
            pass
    win_count.append(win)

win_df = pd.DataFrame({
    'OPP':opp,
    'Win Count':win_count
})

win_df = win_df.sort_values(by=['Win Count'], ascending=False)
```

```python
fig, ax = plt.subplots(1,2,figsize=(15,8))
ax[0].pie(df['Result'].value_counts(), labels=df['Result'].value_counts().index, autopct="%.2f%%")
ax[0].set_title('Win vs Lose', fontsize=15)
ax[1].pie(win_df['Win Count'].values[:5], labels = win_df['OPP'].values[:5],  autopct="%.2f%%")
ax[1].set_title('OPP that lose vs steph', fontsize=15)
plt.show()
```

# Preprocessing

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

result_encoder = encoder.fit(df['Result'])
result_encoder = encoder.transform(df['Result'])

opp_encoder = encoder.fit(df['OPP'])
opp_encoder = encoder.transform(df['OPP'])
```

```python
df['Result'] = result_encoder
df['OPP'] = opp_encoder

df.head()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
df1 = df[['OPP', 'T Score', 'O Score', 'FGM', 'FGA', 'FG%', '3PTM', '3PTA', '3P%', 'FTM', 'FTA', 'FT%', 'PTS', 'Result']]
df1.head()
```

# Split Data

```python
X = df1[['OPP', 'T Score', 'O Score', 'FGM', 'FGA', 'FG%', '3PTM', '3PTA', '3P%', 'FTM', 'FTA', 'FT%', 'PTS']]
y = df['Result']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

# Machine Learning Prediction

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

l_reg = LogisticRegression()
l_reg.fit(X_train, y_train)
```

```python
print('Training Score: ', round(l_reg.score(X_train,y_train)*100,2),'%')
print('Testing Score: ', round(l_reg.score(X_test,y_test)*100,2))
```

```python
y_predict=l_reg.predict(X_test)
y_predict
```

```python
# from sklearn import metrics
# acc=metrics.r2_score(y_test,y_predict)
# print("Accuracy Score of Model: ",round(acc*100,2),'%')

from sklearn.metrics import classification_report

print(classification_report(y_predict, y_test))
```

## SVR

```python
from sklearn.svm import SVC

svr = SVC(C=1.0)
svr.fit(X_train, y_train)
```

```python
print('Training Score: ', round(svr.score(X_train,y_train)*100,2),'%')
print('Testing Score: ', round(svr.score(X_test,y_test)*100,2))
```

```python
y_predict_svr=l_reg.predict(X_test)
y_predict_svr
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_predict_svr, y_test))
```

## KNeighborsRegressor

```python
from sklearn.neighbors import KNeighborsClassifier

knr = KNeighborsClassifier(n_neighbors=5)
knr.fit(X_train, y_train)

print('Training Score: ', round(knr.score(X_train,y_train)*100,2),'%')
print('Testing Score: ', round(knr.score(X_test,y_test)*100,2), '%')
```

```python
y_predict_knr=knr.predict(X_test)
y_predict_knr
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_predict_knr, y_test))
```
