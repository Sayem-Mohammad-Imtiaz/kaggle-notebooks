```python
import cv2
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import glob
```

```python
train_files = glob.glob('../input/cassava-datasetv2/cassava_datasetv2/train_images/*')
train_filenames = [p.split('/')[-1] for p in train_files]
```

```python
train_filenames[:5]
```

```python
len(train_files)
```

```python
def get_img(filepath):
    im_bgr = cv2.imread(filepath)
    im_rgb = im_bgr[:, :, ::-1]
    return im_rgb
```

```python
i_start = 0
plt.figure(figsize=(14, 14))
for i in range(9):
    filepath = train_files[i_start + i]
    filename = filepath.split('/')[-1]
    plt.subplot(3, 3, i+1).set_title(filename)
    img = get_img(filepath)
    plt.imshow(img)
plt.show()
```
