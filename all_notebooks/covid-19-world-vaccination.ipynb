### **Categorical Variable**


**Categorical Variable : country,iso_code,date,source_name,source_website**

```python
def cat_l_plot(fea):
    var=vacc_df[fea]
    var_value=var.value_counts()
    
    plt.figure(figsize=(20,10))
    plt.bar(var_value.index,var_value)
    plt.xticks(var_value.index,var_value.index.values)
    plt.xlabel(fea)
    plt.ylabel("Frequency")
    plt.title(fea+"-plot")
    plt.show()
    print("{} : \n {}".format(fea,var_value))
```

```python
cat_l=["country","iso_code","date","source_name","source_website"]
for i in cat_l:
    cat_l_plot(i)
```

## **Outlier Detection**
q1=percentile,25                                                                                                                                                     
q3=percentile,75                                                                                                                                                     
ıqr= 1.5*(q3-q1)                                                                                                                                                       
outlier<q1-ıqr & outlier>q3+ıqr


```python
def outlier_dec(df,fea):
    outliers=[]
    
    for i in fea:
        q1= np.percentile(df[i],25)
        q3= np.percentile(df[i],75)
        ıqr= 1.5*(q3-q1)
        outlier_list=df[(df[i]<q1-ıqr) | (df[i]>q3+ıqr)].index
        outliers.extend(outlier_list)
    outliers=Counter(outliers)
    multiple_outliers = list(i for i, v in outliers.items() if v > 2)
    return multiple_outliers
        
```

```python
vacc_df.loc[outlier_dec(vacc_df,["total_vaccinations" ,"people_vaccinated" ,"people_fully_vaccinated", "daily_vaccinations_raw","daily_vaccinations", "total_vaccinations_per_hundred", "people_vaccinated_per_hundred", "people_fully_vaccinated_per_hundred" , "daily_vaccinations_per_million"])]
```

```python
vacc_df = vacc_df.drop(outlier_dec(vacc_df,["total_vaccinations" ,"people_vaccinated" ,"people_fully_vaccinated", "daily_vaccinations_raw","daily_vaccinations", "total_vaccinations_per_hundred", "people_vaccinated_per_hundred", "people_fully_vaccinated_per_hundred" , "daily_vaccinations_per_million"]),axis = 0).reset_index(drop = True)
```
