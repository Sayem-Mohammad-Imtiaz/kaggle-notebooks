```python
import os
import numpy 
import pandas

import folium
```

```python
dataframe = pandas.read_csv('../input/all-uk-active-companies-by-sic-and-geolocated/AllCompanies2.csv')
dataframe = dataframe.drop_duplicates(subset=['Latitude', 'Longitude'])
dataframe = dataframe[dataframe.SIC != 'None'].reset_index(drop=True)
dataframe
```

```python
most_common_sic = dataframe['SIC'].value_counts()

sic_desc = pandas.read_csv('../input/all-uk-active-companies-by-sic-and-geolocated/SIC.csv', dtype={'SIC Code': str})
sic_desc = pandas.Series(sic_desc['Description'].values,index=sic_desc['SIC Code']).to_dict()

for i, m in enumerate(most_common_sic[:10]):
    print('%s | %s | Count = %s' % (most_common_sic.index[i], sic_desc[most_common_sic.index[i]], m))
```

```python
selected_sic = [
    '62020', # Information technology consultancy activities
    '70229' # Management consultancy activities other than financial management
]
```

```python
dataframe = dataframe[dataframe.SIC.apply(lambda sic: sic in selected_sic)].reset_index(drop=True)
dataframe
```

```python
colormap = {
    '62020': 'red',
    '70229': 'blue'
}
```

```python
sample = dataframe.sample(1000)
```

```python
map_ = folium.Map(location=[sample.iloc[0]['Latitude'], sample.iloc[0]['Longitude']], zoom_start=11)

for coord in sample[['Latitude', 'Longitude', 'SIC']].to_dict(orient='records'):
    folium.CircleMarker(location=[coord['Latitude'], coord['Longitude']], fill_color=colormap[coord['SIC']], radius=10, weight=0, fill_opacity=1).add_to(map_)

map_
```
