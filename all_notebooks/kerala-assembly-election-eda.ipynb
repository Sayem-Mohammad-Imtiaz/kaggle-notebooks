```python
party=election.groupby(['Party'])['Margin %'].sum().sort_values().to_frame()
party
```

```python
plt.figure(figsize=(15,4))
plt.bar(party.index,party['Margin %'],color='pink')
plt.xlabel('Party')
plt.ylabel('Margin %')
plt.xticks(rotation=90)
plt.show()
```
