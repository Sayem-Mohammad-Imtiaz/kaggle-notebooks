## 2.Exploratory Data Analysis <a class="anchor" id="Chapter2"></a>


### 2.1 Plotly<a class="anchor" id="section_2_1"></a>


```python
fig = px.scatter(df, x="TIME", y="Value", hover_name='LOCATION',hover_data=['MEASURE'],color='SUBJECT',title='Meat Consumption Change through the Years')
fig.show()
```

When looking on a global scale we see the largest rise in poultry, which nearly tripled, this could be due to the efficiency of the animal. Poultry production has a few things that could explain this boom. When looking at animal production feed conversion ratio is an important metric. This is the amount of food needed to increase the animal's weight by one. so the lower the ratio the better the animal is at growing. Chickens have a [feed conversion ratio of 1.6](https://en.wikipedia.org/wiki/Feed_conversion_ratio#:~:text=In%20animal%20husbandry%2C%20feed%20conversion,feed%20into%20the%20desired%20output.&text=FCR%20is%20the%20mass%20of,mass%20of%20milk%20or%20meat). This paired with the resource requirements of the small animal make it perfect for large scale and local production. we also see a large growth in pigs which are considered to have a fair to good [conversion ratio of 3.8-4](https://en.wikipedia.org/wiki/Feed_conversion_ratio#:~:text=In%20animal%20husbandry%2C%20feed%20conversion,feed%20into%20the%20desired%20output.&text=FCR%20is%20the%20mass%20of,mass%20of%20milk%20or%20meat). Pigs also can eat from an extremely wide variety of sources making them easier to raise. They have been called a living [garbage disposal](https://www.washingtonian.com/2018/08/27/dc-used-to-feed-its-trash-to-pigs-whyd-we-stop/).

But this data is hard to read with the aggregate groups like BRICS (Brazil, Russia, India, China, and South Africa) OECD (Organisation for Economic Co-operation and Development) so let's separate these graphs. 

```python
df1=df[df['LOCATION'].isin(['WLD','BRICS','OECD','EU28'])]
```

```python
fig = px.scatter(df1, x="TIME", y="Value", hover_name='SUBJECT',hover_data=['MEASURE'],color='SUBJECT',symbol='LOCATION',
                title='Aggrogate Measurments of Meat Production')
fig.show()
```

I find this graph very interesting. We can start to isolate some interesting trends. if we highlight just the world columns we can see as a whole we are becoming much better fed even with the rising population our overall access to meat has grown tremendously. Yet this doesn't paint the entire picture. Just because the overall production is increasing doesn't mean that the resource is being equitably distributed. 

```python
df2=df[df['LOCATION'].isin(['WLD','BRICS','OECD','EU28'])==False]
```

```python
fig = px.scatter(df2, x="TIME", y="Value",symbol='SUBJECT',hover_data=['MEASURE'],color='LOCATION',hover_name='SUBJECT',
                title='Meat Production by Country and Type')
fig.show()
```

For those who have not used Plotly before you can double click on items in the legend to isolate them and repeat to return to normal. this graph shows two measurements per line in the legend. Say you isolate a variable. You will notice two dotted lines are shown the bottom is in KG per capita this is how much of that eat source people get per year. and the total output in thousands of tons or 'THND_TONNE'.
