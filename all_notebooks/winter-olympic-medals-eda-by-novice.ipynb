```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../input/winter-olympic-medals-1924-2018/Winter_Olympic_Medals.csv')
```

```python
df_copy = df.copy()
df_copy.head()
```

```python
df_copy.dtypes
```

```python
for col in df.select_dtypes('object').columns:
    print(col, '- # unique values:', df[col].nunique())
```

```python
df_copy['Country_Name'].value_counts()
```

```python
df_copy['total'] = df_copy['Gold'] + df_copy['Silver'] + df_copy['Bronze']
df_copy
```

<h1><font size="5">TOP 10 countries that got many medals</font></h1>

```python
top = df.groupby(by=['Country_Name']).sum().groupby(level=[0]).cumsum().drop('Year',axis=1)
top5 = top.sort_values(by="Gold",ascending=False).head(10)

top5.plot(kind="bar",figsize=(20, 10))

plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

<h1><font size="5">Transition of each medals</font></h1>

```python
t = df_copy.groupby(['Year']).agg({'Gold':'sum','Silver':'sum','Bronze':'sum'}).reset_index()

ax1 = t.plot(x="Year", y="Gold",figsize=(10,5))
t.plot(x="Year", y="Silver", ax = ax1)
t.plot(x="Year", y="Bronze", ax = ax1)
```

```python
df_copy.groupby(['Year']).sum()['total'].plot.line(figsize=(10,5))
```

<h1><font size="5">It's total medals sorted 10 countries in order of Country Name</font></h1>

```python
aqua = df_copy.groupby(['Country_Name']).total.sum()
top_10_aqua = aqua.sort_values(ascending = False)[:10]

plt.figure(figsize=(10,5))
top_10_aqua_bar = top_10_aqua.plot.bar()
for p in top_10_aqua_bar.patches:
    top_10_aqua_bar.annotate(f'{int(p.get_height())}', 
                        (p.get_x() + p.get_width() / 2.0, 
                         p.get_height()), 
                         ha = 'center', 
                         va = 'center', 
                         xytext = (0, 4),
                         textcoords = 'offset points')
plt.xticks(rotation=45)
plt.xlabel('Medals earned in order of Country Name  (top 10 countries)')
```

<h1><font size="5">It's total medals by Sorted 10 countries in order of Host country</font></h1>

```python
Host_city = df_copy.groupby(['Host_country']).sum()['total']
Host_city = Host_city.sort_values(ascending = False)[:10]
plt.figure(figsize=(10,5))
Host_city_bar = Host_city.plot.bar()
for p in Host_city_bar.patches:
    Host_city_bar.annotate(f'{int(p.get_height())}', 
                        (p.get_x() + p.get_width() / 2.0, 
                         p.get_height()), 
                         ha = 'center', 
                         va = 'center', 
                         xytext = (0, 4),
                         textcoords = 'offset points')
plt.xticks(rotation=45)
plt.xlabel('Medals earned in host country (top 10 countries)')
```
