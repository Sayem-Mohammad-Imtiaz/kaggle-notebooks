## Training and Testing Data

Now that we have explored the data, let's split it into training and testing sets.
#Set a variable X equal to the numerical features of the customers and a variable Y equal to the 'Yearly Amount Spent' column.

```python
x = df[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
y = df['Yearly Amount Spent']
```

Use Cross validation from sklearn to split the data into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=101)
```

## Training the Model 

```python
import sklearn.linear_model
from sklearn.linear_model import LinearRegression 
```

```python
lm = LinearRegression()
```

Fitting the train data on linear regression model

```python
lm.fit(x_train,y_train)
```

Checking the coefficient of the model

```python
lm.coef_
```

# Predicting on the test data
On our Model Fit & Evolve the performance by prediciting on test values

```python
pred = lm.predict(x_test)
```

Creating a Scatter plot between Predicted value and Acutal values

```python
sns.scatterplot(x=pred,y=y_test)
plt.xlabel('Predicted Values')
plt.ylabel('Actual Values')
```

# Evaluating our Model

Let's calculate residual sum of squares and the variance score

Calculating MAE, MSE & RMSE

```python
from sklearn import metrics
```

```python
print('MAE',metrics.mean_absolute_error(y_test,pred))
print('MSE',metrics.mean_squared_error(y_test,pred))
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test,pred)))
```

```python
# R SQU
metrics.explained_variance_score(y_test,pred)
```

# Residuals

```python
sns.distplot((y_test-pred),bins=50)
```
