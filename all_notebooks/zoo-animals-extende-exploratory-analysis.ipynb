```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
zoo2 = pd.read_csv('../input/zoo-animals-extended-dataset/zoo2.csv')
zoo3 = pd.read_csv('../input/zoo-animals-extended-dataset/zoo3.csv')
```

```python
fig, ax = plt.subplots(5,3, figsize=(20,18))
sns.countplot(zoo2['hair'], ax=ax[0,0])
sns.countplot(zoo2['feathers'], ax=ax[0,1])
sns.countplot(zoo2['eggs'], ax=ax[0,2])
sns.countplot(zoo2['milk'], ax=ax[1,0])
sns.countplot(zoo2['airborne'], ax=ax[1,1])
sns.countplot(zoo2['predator'], ax=ax[1,2])
sns.countplot(zoo2['toothed'], ax=ax[2,0])
sns.countplot(zoo2['backbone'], ax=ax[2,1])
sns.countplot(zoo2['breathes'], ax=ax[2,2])
sns.countplot(zoo2['venomous'], ax=ax[3,0])
sns.countplot(zoo2['fins'], ax=ax[3,1])
sns.countplot(zoo2['legs'], ax=ax[3,2])
sns.countplot(zoo2['tail'], ax=ax[4,0])
sns.countplot(zoo2['catsize'], ax=ax[4,1])
sns.countplot(zoo2['class_type'], ax=ax[4,2])
```
