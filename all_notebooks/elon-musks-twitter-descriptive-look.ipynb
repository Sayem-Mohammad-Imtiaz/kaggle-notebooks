```python
import numpy as np 
import pandas as pd 
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv("../input/elon-musks-tweets/data_elonmusk.csv", encoding = "latin1")
df.head()
```

```python
df["User"].unique()
```

First, let us do some data preprocessing.

There is only one "object" value in User, and row ID seems meaningless, we can drop them, and parse date time, precision units in seconds is way too accurate. 

```python
df.drop(["row ID", "User"],axis =1, inplace= True)
df.head()
```

```python
df.Time = pd.to_datetime(df.Time)
df["year"] = df.Time.dt.year
df["hour"] = df.Time.dt.hour
df.hour
```

After parsing the datetime, I extract year and hour of the day, to see when he is more likely to use twitter

```python
def fun(hour):
    if hour<=12 & hour >= 7:     
        hour = "morning"
    elif hour>12 & hour<= 20:
        hour = "afternon"
    else:
        hour = "night"
    return hour

df.hour = df.hour.apply(lambda x:fun(x) )
df
```

```python
import matplotlib.pyplot as plt
```

```python
counts = df.hour.value_counts()
plt.bar(counts.index,counts)
```

```python
year_counts = df.year.value_counts()
plt.bar(year_counts.index,year_counts)
```

As the plots show, Elon likes to post in the afternoon (between 12 and 8pm), the statistics shows that most of the twitter was posted in the afternoon, morning is least likely.
In perspective of year, twitter useage is ascending in general, 2013 is the first time SpaceX successfully recycled the rocket. That is why he posted significantly different from 2012.
