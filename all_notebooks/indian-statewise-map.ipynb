```python
import numpy as np
from numpy import array
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.basemap import Basemap

from numpy import array
from matplotlib import cm
from matplotlib.dates import date2num
from mpl_toolkits.basemap import Basemap

from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.colors import Normalize
```

```python
df=pd.read_csv('../input/ataljalyojana/Atal Jal 31 March 2021 .xlsx - Sheet1.csv')
```

```python
df['Well Depth']=df['Well Depth'].replace(np.nan, 0)
```

```python
df['State']=df['State'].str.lower()
```

```python
fig=plt.figure(figsize=(20,20))
map=Basemap(width=1200000,height=900000,llcrnrlon=67,llcrnrlat=6, urcrnrlon=97, urcrnrlat=37, projection='lcc', lat_0=28, lon_0=77.)
map.drawmapboundary()
#map.drawcounties()
map.drawcoastlines(linewidth=4, color='blue')
map.drawcountries(linewidth=4, color='red')
map.readshapefile('../input/india-shape/India_Districts_ADM2_GADM', 'india-shape')
map.drawlsmask(land_color='0.8', ocean_color='#7777ff', lakes=True)
map.shadedrelief()

lg=array(df['Longitude'])
lt=array(df['Latitude'])
pt=array(df['Well Depth'])

x, y = map(lg, lt)
well_depth = df['Well Depth'].apply(lambda x: int(x))
plt.scatter(x, y,s=well_depth,marker='*', cmap=cm.Dark2, linewidths=3)
plt.title("Map of India with Longitute and Latitude ")

plt.show()
```
