<h1>If Mushroom Smells Nothing Than It Is Edible</h1>

This statement is surprisingly true for 88% of these 8124 mushrooms (23 species of gilled mushrooms in the Agaricus and Lepiota Family). Let me show you!

```python
import pandas as pd
```

```python
data = pd.read_csv("/kaggle/input/mushroom-classification/mushrooms.csv")
```

```python
data = pd.get_dummies(data)
```

```python
corr = data.corr()
```

You can see below there is a strong positive correlation between edible mushrooms and none smelling mushrooms.

```python
corr["class_e"].sort_values(ascending=False)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_true = data["class_e"]
y_pred = data["odor_n"]
```

```python
accuracy_score(y_true, y_pred)
```

See, If you only eat none smelling mushrooms, you are 88 percent likely to survive!
