```python
from sklearn.utils import shuffle
df = pd.read_csv("/kaggle/input/toydatasets/concertriccir2.csv")

df = shuffle(df, random_state = 19)
df = df.reset_index()
```

```python
X = df.iloc[:, 1:3]
display(X)
y = df.iloc[:, 3:]
display(y)
```

```python
sns.scatterplot(data = df.iloc[:400, :], x = 'X', y = 'Y', hue = "class")
```

```python
X_train = X.iloc[:400, :]
X_val = X.iloc[400:450, :]
X_test = X.iloc[450:, :]
y_train = y.iloc[:400, :]
y_val = y.iloc[400:450, :]
y_test = y.iloc[450:, :]
display(X_test)
```

```python
from sklearn import neighbors, metrics

model = sklearn.neighbors.KNeighborsClassifier(5)
model.fit(X_train, y_train)

y_pred = model.predict(X_train)
print(sklearn.metrics.accuracy_score(y_train, y_pred))
```

```python
y_pred_val = model.predict(X_val)
print(sklearn.metrics.accuracy_score(y_val, y_pred_val))
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x = X_train.loc[:, "X"], y = X_train.loc[:, "Y"], hue = y_train.loc[:, "class"], style = y_pred)
```

```python
sns.scatterplot(x = X_val.loc[:, "X"], y = X_val.loc[:, "Y"], hue = y_val.loc[:, "class"], style = y_pred_val)
```

```python
k_values = []
accuracies = []
for i in range(1, 15):
    model = sklearn.neighbors.KNeighborsClassifier(i)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_train)
    #print(sklearn.metrics.accuracy_score(y_train, y_pred))
    k_values.append(i)
    y_pred_val = model.predict(X_val)
    accuracies.append(sklearn.metrics.accuracy_score(y_val, y_pred_val))
```

```python
sns.lineplot(x = k_values, y = accuracies)
```

```python
model = sklearn.neighbors.KNeighborsClassifier(5)
model.fit(X_train, y_train)

y_pred_test = model.predict(X_test)
print(sklearn.metrics.accuracy_score(y_test, y_pred_test))
```
