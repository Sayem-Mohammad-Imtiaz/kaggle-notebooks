```python
#SETUP YOUR NETBOOK


import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
rio_data = pd.read_csv("../input/temperature-timeseries-for-some-brazilian-cities/station_rio.csv",index_col="YEAR")
rio_data.head()
```

```python
graphe = rio_data[rio_data.columns[0:12]]

# Set the width and height of the figure
plt.figure(figsize=(15,9))

# Line chart 
sns.lineplot( data=graphe[(graphe < 100)], dashes=False)
```

```python
sns.distplot(graphe.query('JAN < 100')['JAN'], bins=20)
```

```python
plt.figure(figsize=(15,9))
sns.barplot(x=graphe[(graphe < 100)].index, y=rio_data[(rio_data < 100)]["MAR"])
```
