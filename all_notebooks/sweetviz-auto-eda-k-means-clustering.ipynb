# Preprocessing

```python
scaler = StandardScaler()
data = scaler.fit_transform(df)
```

# Model

```python
n_clusters = 10
distortions = []
ticks=[]
for n in range(1, n_clusters):
    kmeans = KMeans(n_clusters=n, random_state=42)
    kmeans.fit(data)
    distortions.append(kmeans.inertia_)
    ticks.append(n)
```

```python
plt.figure(figsize=(16, 8))
plt.plot(ticks, distortions)
```

```python
elbow_point = 3
```

```python
kmeans = KMeans(n_clusters=elbow_point)
kmeans.fit(data)
```

# Final Visualization with the Help of PCA

```python
pca = PCA(2)
data_cluster = data.copy()
data_cluster = pca.fit_transform(data_cluster)
```

```python
plt.scatter(data_cluster[:, 0], data_cluster[:, 1], c=kmeans.labels_)
```
