```python
import pandas as pd
import re 
import nltk
```

```python
df= pd.read_csv('../input/all-trumps-twitter-insults-20152021/trump_insult_tweets_2014_to_2021.csv')
```

```python
df
```

```python
train_t=df['tweet']
```

```python
train=[]
for i in train_t:
  text=re.sub(r"[\W]"," ",i[10:].lower())
  text1=re.sub(r'[^a-z0-1\s]','',text)
  train.append(text1) 
```

```python
from nltk.tokenize import word_tokenize, sent_tokenize
nltk.download('punkt')
```

```python
list =[]
for i in train:
  list.append(word_tokenize(i))
list
```

```python
nltk.download('stopwords')
stopwords=nltk.corpus.stopwords.words('english')
words=[]
for i in list:
    i=[a for a in i if a not in stopwords]
    words.append(i)
```

```python
words
```

```python
from gensim import corpora
dictinory = corpora.Dictionary(words)
```

```python
inp_corpus = [dictinory.doc2bow(word,allow_update=True) for word in words]
```

```python
import gensim
model1 =gensim.models.ldamodel.LdaModel(inp_corpus,num_topics=3,id2word=dictinory)
```

```python
topics=model1.print_topics()
topics
```
