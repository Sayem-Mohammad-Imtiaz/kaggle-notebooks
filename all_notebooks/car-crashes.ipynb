```python
dataset_path = '../input/car-crashes-severity-prediction'

df = pd.read_csv(os.path.join(dataset_path, 'train.csv'))

print("The shape of the dataset is {}.\n\n".format(df.shape))

df.head()
```

```python
df["Bump"] = df["Bump"].astype(int)
df["Crossing"] = df["Crossing"].astype(int)
df["Give_Way"] = df["Give_Way"].astype(int)
df["Junction"] = df["Junction"].astype(int)
df["No_Exit"] = df["No_Exit"].astype(int)
df["Railway"] = df["Railway"].astype(int)
df["Roundabout"] = df["Roundabout"].astype(int)
df["Stop"] = df["Stop"].astype(int)
df["Amenity"] = df["Amenity"].astype(int)
df.head()
```

```python
dataset_path = '../input/car-crashes-severity-prediction'

df2 = pd.read_csv(os.path.join(dataset_path, 'weather-sfcsv.csv'))

print("The shape of the dataset is {}.\n\n".format(df.shape))
df2.head()
```

```python
time=[]
d=list(df2["Day"])
m=list(df2["Month"])
y=list(df2["Year"])
for i in range(len(y)):
    s=str(y[i])+"-"+str(m[i])+"-"+str(d[i])
    time.append(s)
del df2['Day']
del df2['Year']
del df2['Month']
del df2['Hour']

df2["Time"]=time
df2
```

```python
#print("The shape of the Weather before removing dublicate is {}.\n\n".format(df_weather.shape))
df2=df2.drop_duplicates(['Time'], keep='last')
print("The shape of the Weather After removing dublicate is {}.\n\n".format(df2.shape))
df2
```

```python
dataset_path = '../input/inputcarcrashesseverityprediction'

df3 = pd.read_csv(os.path.join(dataset_path, 'Holidays.csv'))

print("The shape of the dataset is {}.\n\n".format(df3.shape))
df3.rename(columns = {'Time':'date'}, inplace = True)
df3.head()
```

```python
df3.rename(columns = {'date':'Time'}, inplace = True)
df3
```

```python
merged=df.merge(df2,how="left",on=['Time'])
merged
```

```python
final_merge=merged.merge(df3,how="left",on=['Time'])
final_merge
```

```python
del final_merge["Give_Way"]
del final_merge["description"]
del final_merge["Crossing"]
del final_merge["Weather_Condition"]
```

```python
final_merge
```

```python
final_merge['Visibility(mi)'].fillna((final_merge['Visibility(mi)'].mean()), inplace=True)
final_merge['Precipitation(in)'].fillna((final_merge['Precipitation(in)'].mean()), inplace=True)
final_merge['Wind_Chill(F)'].fillna((final_merge['Wind_Chill(F)'].mean()), inplace=True)
final_merge['Temperature(F)'].fillna((final_merge['Temperature(F)'].mean()), inplace=True)
final_merge['Humidity(%)'].fillna((final_merge['Humidity(%)'].mean()), inplace=True)
final_merge['Wind_Speed(mph)'].fillna((final_merge['Wind_Speed(mph)'].mean()), inplace=True)
```

```python
from sklearn.model_selection import train_test_split

train_df, val_df = train_test_split(final_merge, test_size=0.2, random_state=42) # Try adding `stratify` here

X_train = train_df.drop(columns=['ID', 'Severity'])
y_train = train_df['Severity']

X_val = val_df.drop(columns=['ID', 'Severity'])
y_val = val_df['Severity']
```

```python
# This cell is used to select the numerical features. IT SHOULD BE REMOVED AS YOU DO YOUR WORK.
X_train = X_train[['Lat', 'Lng', 'Distance(mi)','Humidity(%)','Wind_Speed(mph)','Junction','Stop','Amenity','Precipitation(in)','Temperature(F)','Visibility(mi)']]
X_val = X_val    [['Lat', 'Lng', 'Distance(mi)','Humidity(%)','Wind_Speed(mph)','Junction','Stop','Amenity','Precipitation(in)','Temperature(F)','Visibility(mi)']]
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create an instance of the classifier
classifier = RandomForestClassifier(max_depth=2, random_state=0)

# Train the classifier
classifier = classifier.fit(X_train, y_train)
```

```python
print("The accuracy of the classifier on the validation set is ", (classifier.score(X_val, y_val)))
```
