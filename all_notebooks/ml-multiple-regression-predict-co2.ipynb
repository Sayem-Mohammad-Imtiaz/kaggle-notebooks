# Machine Learning - Multiple Regression

Multiple regression is like linear regression, but with more than one independent value, meaning that we try to predict a value based on two or more variables.
We can predict the CO2 emission of a car based on the size of the engine, but with multiple regression we can throw in more variables, like the weight of the car, to make the prediction more accurate Created this Notebook with education purposes




```python
import pandas
from sklearn import linear_model

df = pandas.read_csv("/kaggle/input/ml-multiple-regresion-cars-co2/cars.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

#predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3:
predictedCO2 = regr.predict([[2300, 1300]])

print(predictedCO2)
```

We predicted that the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3: are 107.2087328 g

The result array represents the coefficient values of weight and volume:

```python
print(regr.coef_)
```
