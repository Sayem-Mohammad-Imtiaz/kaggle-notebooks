```python
node1 = tf.constant(3.0, tf.float32)
node2 = tf.constant(4.0) # also tf.float32 implicity
print(node1, node2)
```

```python
# run the computation graph inside a Session to evaluate node1 and node2
sess = tf.Session()
print(sess.run([node1, node2]))
```

```python
node3 = tf.add(node1, node2)
print("node3: ", node3)
print("sess.run(node3): ", sess.run(node3))
```

```python
a = tf.placeholder(tf.float32)
b = tf.placeholder(tf.float32)
adder_node = a + b # + provides a shortcut for tf.add(a, b)
print(sess.run(adder_node, {a: 3, b: 4.5}))
print(sess.run(adder_node, {a: [1, 3], b: [-12, -21]}))
```

```python
add_and_triple = adder_node * 3
print(sess.run(add_and_triple, {a: [1,3], b: [-12, -21]}))
```

```python
W = tf.Variable([.3], tf.float32)
b = tf.Variable([-.3], tf.float32)
x = tf.placeholder(tf.float32)
linear_model = W * x + b
# print(sess.run(W)) would give an error because the variables are not initialized when you call tf.Variable

init = tf.global_variables_initializer()
sess.run(init)

print(W)
print(sess.run(W))

print(sess.run(linear_model, {x: [1,2,3,4]}))
```

```python
y = tf.placeholder(tf.float32)
squared_deltas = (linear_model - y) * (linear_model - y)
loss = tf.reduce_sum(squared_deltas)
print(sess.run(squared_deltas, {x:[1,2,3,4], y:[0,-1,-2,-3]}))
print(sess.run(loss, {x:[1,2,3,4], y:[0,-1,-2,-3]}))
```

```python
# Assign the perfect value of W and b
# tf.Variable can be re-assigned
fixW = tf.assign(W, [-1.])
fixb = tf.assign(b, [1])
sess.run([fixW, fixb])
print(sess.run(loss, {x:[1,2,3,4], y:[0,-1,-2,-3]}))
```

## A higher level API -- tf.train ##

```python
optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(loss)

sess.run(init) # reset values to incorrect defaults.
for i in range(10):
    sess.run(train, {x:[1,2,3,4], y:[0,-1,-2,-3]})
print(sess.run([W,b]))

sess.run(init) # reset values to incorrect defaults.
for i in range(100):
    sess.run(train, {x:[1,2,3,4], y:[0,-1,-2,-3]})
print(sess.run([W,b]))

sess.run(init) # reset values to incorrect defaults.
for i in range(1000):
    sess.run(train, {x:[1,2,3,4], y:[0,-1,-2,-3]})
print(sess.run([W,b]))
```

```python
# full program
import numpy as np # linear algebra
import tensorflow as tf

# define the variables and placeholder
x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
W = tf.Variable(.3, tf.float32)
b = tf.Variable(-.3, tf.float32)

# define the training data
x_train = [1,2,3,4]
y_train = [0,-1,-2,-3]

# define the functions
sess = tf.Session()
linear_model = W * x + b
loss = tf.reduce_sum(tf.square(linear_model - y))
init = tf.global_variables_initializer()
optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(loss)

# training loop
sess.run(init) # reset value to wrong
for i in range(1000):
    sess.run(train, {x:x_train, y:y_train})

# evaluate training accuracy
curr_W, curr_b, curr_loss = sess.run([W, b, loss], {x:x_train, y:y_train})
print("W: %s b: %s loss: %s "%(curr_W, curr_b, curr_loss))
```
