Creating the heatmap

```python
import seaborn as sns
from matplotlib.pyplot import figure
figure(figsize= (13,12))
sns.heatmap(dataset.corr(),annot= True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score, recall_score, f1_score, precision_score
```

Preprocess the values and create the dataset

```python
dataset["age"]= np.log(dataset.age)
dataset["trtbps"]= np.log(dataset.trtbps)
dataset["chol"]= np.log(dataset.chol)
dataset["thalachh"]= np.log(dataset.thalachh)
y = dataset['output'].values
X = dataset.iloc[:,:-1].values
```

Train/Test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

Now, let's use Classification models to predict heart attack

1. Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
RFclassifier = RandomForestClassifier(n_estimators = 50, criterion = 'entropy', random_state = 0)
RFclassifier.fit(X_train, y_train)
```

Visualizing the confusion matrix

```python
y_pred_RF = RFclassifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred_RF)
figure(figsize= (8,6))
sns.heatmap(cm, annot = True)
print('Confusion matrix: ')
```

Scores

```python
print('ACCURACY:', accuracy_score(y_test, y_pred_RF))

print('How often is the classifier incorrect?: ', 1-accuracy_score(y_test, y_pred_RF))

print('Sensitivity or Recall - When the actual value is positive, how often does it predict correct answer?',recall_score(y_test, y_pred_RF))

print('Precision - when a positive value is predicted, how often is the prediction correct?', precision_score(y_test, y_pred_RF))

print('F1: ', f1_score(y_test, y_pred_RF))
```

2. Support Vector Classifier

```python
from sklearn.svm import SVC
SVCclassifier = SVC(kernel = 'rbf', random_state = 0)
SVCclassifier.fit(X_train, y_train)
```

Visualizing the confusion matrix

```python
y_pred_SVC = SVCclassifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred_SVC)
figure(figsize= (8,6))
sns.heatmap(cm, annot = True)
print('Confusion matrix: ')
```

Scores

```python
print('ACCURACY:', accuracy_score(y_test, y_pred_SVC))

print('How often is the classifier incorrect?: ', 1-accuracy_score(y_test, y_pred_SVC))

print('Sensitivity or Recall - When the actual value is positive, how often does it predict correct answer?',recall_score(y_test, y_pred_SVC))

print('Precision - when a positive value is predicted, how often is the prediction correct?', precision_score(y_test, y_pred_SVC))

print('F1: ', f1_score(y_test, y_pred_SVC))
```

3. KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier
KNNclassifier= KNeighborsClassifier(n_neighbors= 50)
KNNclassifier.fit(X_train,y_train)
```

Visualizing the confusion matrix

```python
y_pred_KNN = KNNclassifier.predict(X_test)
cm = confusion_matrix(y_test, y_pred_KNN)
figure(figsize= (8,6))
sns.heatmap(cm, annot = True)
print('Confusion matrix: ')
```

Scores

```python
print('ACCURACY OF K-nearest neighbors:', accuracy_score(y_test, y_pred_KNN))

print('How often is the classifier incorrect?: ', 1-accuracy_score(y_test, y_pred_KNN))

print('Sensitivity or Recall - When the actual value is positive, how often does it predict correct answer?',recall_score(y_test, y_pred_KNN))

print('Precision - when a positive value is predicted, how often is the prediction correct?', precision_score(y_test, y_pred_KNN))

print('F1: ', f1_score(y_test, y_pred_KNN))
```
