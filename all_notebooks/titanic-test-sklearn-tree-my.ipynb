```python
import pandas as pd
import numpy as np
train = pd.read_csv("../input/titanic/train.csv")
test = pd.read_csv("../input/titanic/test.csv")

test.head(10)
```

```python
test_shape = test.shape
train_shape = train.shape

print(test_shape)
print(train_shape)
```

```python
def kesson_table(df):
    null_val = df.isnull().sum()
    percent = 100 * null_val/len(df)
    kesson_table = pd.concat([null_val,percent],axis=1)
    kesson_table_ren_columns = kesson_table.rename(columns={0:'欠損数',1:'%'})
    return kesson_table_ren_columns

kesson_table(train)
```

```python
kesson_table(test)
```

```python
train["Age"] = train["Age"].fillna(train["Age"].median())
train["Embarked"] = train["Embarked"].fillna("S")
kesson_table(train)
```

```python
test["Age"] = test["Age"].fillna(test["Age"].median())
test["Fare"] = test["Fare"].fillna(test["Fare"].median())
kesson_table(test)
```

```python
#ここの文法全くわからん
train["Sex"][train["Sex"] == "male"] = 0
train["Sex"][train["Sex"] == "female"] = 1
train["Embarked"][train["Embarked"] == "S" ] = 0
train["Embarked"][train["Embarked"] == "C" ] = 1
train["Embarked"][train["Embarked"] == "Q"] = 2
 
train.head(10)
```

```python
test["Sex"][test["Sex"] == "male"] = 0
test["Sex"][test["Sex"] == "female"] = 1
test["Embarked"][test["Embarked"] == "S" ] = 0
test["Embarked"][test["Embarked"] == "C" ] = 1
test["Embarked"][test["Embarked"] == "Q"] = 2
 
test.head(10)
```

```python
#ここらへんは要勉強
from sklearn import tree
#学習する
target = train["Survived"].values
features = train[["Pclass","Sex","Age","SibSp","Parch","Fare"]].values

max_depth = 10
min_samples_split = 0.01

my_tree = tree.DecisionTreeClassifier(max_depth = max_depth, min_samples_split = min_samples_split)
my_tree = my_tree.fit(features, target)

test_features = test[["Pclass","Sex","Age","SibSp","Parch","Fare"]].values

my_prediction = my_tree.predict(test_features)

my_prediction.shape
```

```python
PassengerId = np.array(test["PassengerId"]).astype(int)

my_solution_tree = pd.DataFrame(my_prediction, PassengerId, columns = ["Survived"]) #データフレームの形にする　列名のはどうやって決めてる？（片方だけ？）
my_solution_tree.to_csv("my_tree_210417.csv", index_label = ["PassengerId"]) 
```
