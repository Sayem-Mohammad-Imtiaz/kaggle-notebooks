```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
buffet = pd.read_csv('../input/brka-and-ff-factors/buffet.csv')
buffet.describe()

buffet.tail()
```

```python
buffet['BUF-Mkt'] = buffet['BUF-RF'] - buffet['Mkt-RF']
print('The monthly active return is %6.4f'%(buffet[['BUF-Mkt']].mean()) )
```

```python
Sharpe_Ratio = np.sqrt(12)*(buffet['buffet'].mean() - buffet['RF'].mean()) / buffet['buffet'].std()
print('Sharpe Ratio is %6.4f' %(Sharpe_Ratio))
```

```python
import statsmodels.api as sm # import statsmodels 
X = buffet[["Mkt-RF"]] ## X usually means our input variables (or independent variables)
y = buffet[["BUF-RF"]] ## Y usually means our output/dependent variable
X = sm.add_constant(X) ## let's add an intercept (beta_0) to our model

# Note the difference in argument order
model = sm.OLS(y, X).fit() ## sm.OLS(output, input)

# Print out the statistics

print('Jensen\'s Alpha is %6.4f' %(model.params['const']) )

model.summary()
```

```python
#Information ratio
IR = buffet['BUF-RF'].mean() / buffet['BUF-RF'].std()
print ('Information Ration is %6.4f' %(IR)) 
```

```python
#Fama and French Three-Factor Alpha
Xa = buffet[["Mkt-RF","SMB","HML"]] ## X usually means our input variables (or independent variables)
ya = buffet[["BUF-RF"]] ## Y usually means our output/dependent variable
Xa = sm.add_constant(Xa) ## let's add an intercept (beta_0) to our model

# Note the difference in argument order
model = sm.OLS(ya, Xa).fit() ## sm.OLS(output, input)

# Print out the statistics

print('Fama and French Three-Factor Alpha is %6.4f' %(model.params['const']) )

model.summary()
```

```python
#Fama and French Five-Factor Alpha
Xb = buffet[["Mkt-RF","SMB","HML","RMW","CMA"]] ## X usually means our input variables (or independent variables)
yb = buffet[["BUF-RF"]] ## Y usually means our output/dependent variable
Xb = sm.add_constant(Xb) ## let's add an intercept (beta_0) to our model

# Note the difference in argument order
model = sm.OLS(yb, Xb).fit() ## sm.OLS(output, input)

# Print out the statistics

print('Fama and French Five-Factor Alpha is %6.4f' %(model.params['const']) )

model.summary()
```
