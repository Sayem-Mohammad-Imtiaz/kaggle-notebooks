```python
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('../input/infant-mortality/InfantMortalityRate.csv')
```

```python
df = df.dropna(axis = 0)
```

```python
df_clean = df.drop_duplicates().dropna().astype({"Year": "int", "Gender" : "category"})
```

```python
df_clean=df_clean[df_clean.Gender != 'Total']
```

```python
df_clean = df_clean[df_clean.Country != 'World']
```

```python

sns.displot(data=df_clean['Infant Mortality Rate'])
```

```python
top5 = pd.DataFrame(df_clean.groupby('Country')['Infant Mortality Rate'].sum().sort_values(ascending=False).head())
```

```python
top5
```

```python
top5_cntry = ['sub-Saharan Africa','Sierra Leone','Central African Republic','Somalia','Nigeria']
```

```python
sns.barplot(data = top5,x='Infant Mortality Rate',y = top5_cntry)
plt.title('Top 5 countries With Highest Female and Male Infant Mortality over the years')
plt.show()
```

```python
bottom5 = pd.DataFrame(df_clean.groupby('Country')['Infant Mortality Rate'].sum().sort_values(ascending=False)).tail()
```

```python
bottom5
```

```python
bottom5_country = ['Finland','Slovenia','Japan','San Marino','Iceland']
```

```python
sns.barplot(data = bottom5,x = 'Infant Mortality Rate',y = bottom5_country )
plt.title('bottom 5 countries With Highest Female and Male Infant Mortality over the years')
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go


import cufflinks as cf
cf.go_offline()
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
px.line(df_clean,x = 'Year',y = 'Infant Mortality Rate',color = 'Country')
```

```python
fig3 = px.strip(df_clean.reset_index(), y="Infant Mortality Rate", x="Year", color="Gender", hover_data=df_clean.columns)
fig3.add_trace(go.Scatter(x=df_clean.groupby('Year').max().index, y=df_clean.groupby('Year').max()['Infant Mortality Rate'], mode='lines', line={'width': 4, 'color': 'black'}, name='Maximum'))
fig3.add_trace(go.Scatter(x=df_clean.groupby('Year').min().index, y=df_clean.groupby('Year').min()['Infant Mortality Rate'], mode='lines', line={'width': 4, 'color': 'green'}, name='Minimum'))
fig3.add_trace(go.Scatter(x=df_clean.groupby('Year').mean().index, y=df_clean.groupby('Year').mean()['Infant Mortality Rate'], mode='lines', line={'width': 4, 'color': 'blue'}, name='Mean'))
fig3.show()
```
