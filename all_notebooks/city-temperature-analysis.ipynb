Setting up the notebook by loading in the required libraries:

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider, Button
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

Creating a variable named "climate_filepath" that would store in filepath and another variable named "climate_data" that would read the given filepath:

```python
climate_filepath = "../input/daily-temperature-of-major-cities/city_temperature.csv"

climate_data = pd.read_csv(climate_filepath)

climate_data.head()
```

Next, we want to filter the dataset by a specific city (in this case Delhi) for that we create a variable that would contain the very dataset filtered, "climate_city_data"

But, the new variable contains the values filtered by boolean terms, in other words, it shows only *True* and *False* values. So, we create another variable "new_climate_data" that would store in the original dataset "climate_data" containing only the *True* values from "climate_city_data".

This way we have a new dataset "new_climate_data" filtered:

```python
climate_city_data = climate_data["City"]=="Delhi" # City to Search

new_climate_data = climate_data[climate_city_data]
```

The same principle will also be now used to filter out the month (in this case July) from that new dataset we had created "new_climate_data":

```python
climate_month_data = new_climate_data["Month"]==7  # Month to Search 

new_final = new_climate_data[climate_month_data]
```

..and finally the day in which we want to filter by, from "new_final"(which is filtered by the selected city and month):

```python
climate_day_data = new_final["Day"]==1 # Day to Search

done = new_final[climate_day_data]
```

With this new variable "done" (not very creative), we have a dataframe filtered by the city, month, and day we had selected.

```python
done.head()
```

Now, it is time to visualize the data. We consider the X values to be taken from the "Year" column and the Y values from "AvgTemperature", in this case the Fahrenheit values have been converted to Celsius with this simple operation --> (°F − 32) × 5/9.

The Y Label has also been renamed to "Avg Temperature ºC" and the plot's title is set to "Delhi Temperature Measurements on July by Year".

```python
sns.lineplot(x = done["Year"], y = (done["AvgTemperature"] - 32) * (5/9))
plt.ylabel("Avg Temperature ºC")
plt.title("Delhi Temperature Measurements on July by Year")
```
