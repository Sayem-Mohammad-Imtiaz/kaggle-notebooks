```python
import numpy as np
import pandas as pd
import os
import folium
from folium import plugins
import matplotlib
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data=pd.read_csv('../input/us-hospital-locations/us_hospital_locations.csv')
data
```

```python
data['BEDS'].value_counts()
```

```python
drop=data[data['BEDS']==-999].index.tolist()
data=data.drop(index=data.index[drop],axis=0).reset_index(drop=True)
data
```

```python
print(data.columns.tolist())
```

```python
data2=data[['STATE','CITY','BEDS']]
data2
```

## Total Beds by State

```python
data2a=data2.groupby('STATE',as_index=False).sum()
data2a=data2a.sort_values(['BEDS'],ascending=False)
data2a[0:20]
```

```python
fig = px.bar(data2a[0:30], x='STATE', y='BEDS',title="Total Beds by State")
fig.show()
```

## Total Beds by City

```python
data2b=data2.groupby('CITY',as_index=False).sum()
data2b=data2b.sort_values(['BEDS'],ascending=False)
data2b[0:20]
```

```python
fig = px.bar(data2b[0:40], x='CITY', y='BEDS',title="Total Beds by City")
fig.show()
```

## Hospitals with more than 1,200 beds 

```python
data3=data[data['BEDS']>1200]
data3[['NAME','STATE','CITY','BEDS']].sort_values('BEDS',ascending=False)
```

```python
coor=data3[['LATITUDE', 'LONGITUDE']]
coor
```

```python
shark_map = folium.Map(location=[40,-100],
                            tiles='Stamen Terrain',
                            zoom_start=4.0,min_zoom = 3.0)
```

```python
shark_map.add_child(plugins.HeatMap(coor))
shark_map
```
