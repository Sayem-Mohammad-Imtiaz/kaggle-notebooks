```python
from sklearn.model_selection import train_test_split

X = df_diamond.loc[:,df_diamond.columns != 'price']
y = df_diamond.price
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 48)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(X_train, y_train)

y_pred = reg.predict(X_test)

# store errors
errors = y_pred - y_test
errors.shape
```

```python
from sklearn.model_selection import KFold

cv = KFold(n_splits=10, random_state=1, shuffle=True)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(reg, X, y, scoring='r2', cv=cv, n_jobs=-1)
# report performance
print('Accuracy: %.3f (%.3f)' % (np.mean(scores), np.std(scores)))
```

```python
y_pred = reg.predict(X_test)
result = pd.DataFrame({"prediction": y_pred, "actual_price": y_test})
result.head()
```
