```python
df = pd.read_csv("../input/sms-spam-collection-dataset/spam.csv",encoding='ISO-8859-1')
```

```python
df.head()
```

```python
#droping the unwanted columns
df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'], axis='columns', inplace=True)
```

```python
df.head()
```

```python
df.shape
```

```python
df['v1'].unique()
```

```python
df['v1'].value_counts()
```

```python
#changing the v1 coulmn to 0 and 1
#ham is represented sa 0
#spam is represented as 1


df['v1']=df['v1'].replace('ham', 0)
df['v1']=df['v1'].replace('spam', 1)
```

```python
df['v1'].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.countplot(df['v1'])
```

```python
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer

from nltk.corpus import stopwords
```

```python
import re
ps=PorterStemmer()

corpus_stem=[]
for i in range(len(df)):
    review=re.sub('[^a-zA_Z]',' ',df['v2'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if word not in set(stopwords.words('english'))]
    review=' '.join(review)
    corpus_stem.append(review)
    
corpus_stem
```

```python
lem=WordNetLemmatizer()

corpus_lem=[]
for i in range(len(df)):
    review=re.sub('[^a-zA_Z]',' ',df['v2'][i])
    review=review.lower()
    review=review.split()
    review=[lem.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
    review=' '.join(review)
    corpus_lem.append(review)
    
corpus_lem
```

```python
from sklearn.feature_extraction.text import CountVectorizer
CV=CountVectorizer(max_features=5000)
x_lem=CV.fit_transform(corpus_lem).toarray()
x_lem
```

```python
x_lem.shape
```

```python
x_stem=CV.fit_transform(corpus_stem).toarray()
x_stem
```

```python
x_stem.shape
```

```python
#TF-IDF for stemming

from sklearn.feature_extraction.text import TfidfVectorizer
tf_stem=TfidfVectorizer()
x_tf_stem=tf_stem.fit_transform(corpus_stem)
print(x_tf_stem.shape)
```

```python
#TF-IDF for lemmetizing
tf_lem=TfidfVectorizer()
x_tf_lem=tf_stem.fit_transform(corpus_lem)
print(x_tf_lem.shape)
```

```python
y=df['v1']
y[:5]
```

### **making the model by using countvectorizer and stemming**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(x_stem,y,test_size=0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB().fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
y_pred=model.predict(X_test)
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted value')
plt.ylabel('True value')
```
