```python
# Count the number of folders, one folder per class
d="../input/spectrodatarml201610a/Datas/input/notebook-data-distribution-by-category/Datas/RML2016.10a_spectrog/"
data_dir = pathlib.Path(d)
mod_count = len(list(data_dir.glob('*')))
print(mod_count)
```

```python
# A function that retrieves information about the spectrogram 
def getImInfo(i):
    im_name=i.filename.split('/')[-1]
    modulation=im_name.split('_')[0]
    snr=im_name.split('_')[1].split('.')[0]
    return im_name,i.size,modulation,snr
```

```python
# Display of a random spectrogram image with its relevant information (modulation, snr,..) 
numb=np.random.randint(1000)
QAM64 = list(data_dir.glob('QAM64/*'))
i=PIL.Image.open(str(QAM64[numb]))

spec_im_name, size, modulation, snr = getImInfo(i)
# spec_im_name=i.filename.split('/')[-1]
# modulation=spec_im_name.split('_')[0]
# snr=spec_im_name.split('_')[1].split('.')[0]
print('Name: ',spec_im_name)
print('Size: ',size)
print('Modulation: ',modulation)
print('SNR: ',snr)

i
```

```python
# Define some parameters 
batch_size = 128
img_height = 163
img_width = 223
```

```python
# Creation of the training set, by splitting the data into training and validation. 
# We will use 80% of the images for training and 20% for validation.
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="training",
  seed=1619,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
# Creation of the validation set (20% of the images)
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
  data_dir,
  validation_split=0.2,
  subset="validation",
  seed=1619,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
# The class names of the datasets.
class_names = train_ds.class_names
print(class_names)
```

```python
# Visualize the data
# The first 9 images from the training dataset.

# plt.style.use("dark_background")
fig = plt.figure(figsize=(14, 10))
for images, labels in train_ds.take(1):
  for i in range(9):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")
```
