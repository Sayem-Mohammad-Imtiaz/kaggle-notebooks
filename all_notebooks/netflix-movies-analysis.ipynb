<p1>But, before proceeding with analysis, lets make some changes to the data to handle the missing values</p1>

<p2>We would be replacing all the missing values in the string fields with the phrase "missing", so that when we tweak these fields to best fit our use case, it wont impact our analysis</p2>

```python

netflix.fillna({'country':'empty','director':'empty','cast':'empty','rating':'missing'},inplace=True)
netflix=netflix[netflix.date_added.isna()==False]
netflix.info()
```

<p1>Now that we have handled the missing values, lets do some basic analysis. Lets start with the split of Movies and TV Shows screened in Netflix so far.</p1>

```python
# Split of Movies and TV Shows
netflix.groupby('type')['show_id'].count()
```

<b>We observe that Netflix fancies more movies than TV Shows</b>

<p1>Lets create a new column to have the year of addition of these shows/movies, which should help us plot the same to see the trend</p1>

```python
# Lets create a new column to represnt the year of addition of Movies/TV Shows to the platform

netflix['year_added']=netflix['date_added'].dt.year
netflix=netflix.astype({'year_added':int})
```

```python
netflix.info()
```

<h3>Plot of the count of movies added to Netflix over time</h3>

```python
a=netflix[netflix.type=='Movie']
a.groupby('year_added')['show_id'].count().plot()
```

<h3>Plot of the count of TV Shows added to Netflix over time</h3>

```python
b=netflix[netflix.type=='TV Show']
b.groupby('year_added')['show_id'].count().plot()
```

```python
netflix.groupby('director')['show_id'].count().sort_values(ascending=False)
```

<p1>Lets split the countries where the production happened and the categories and add then as a list in seperate columns</p1>

```python
category=[]
countries=[]
for i in netflix['listed_in']:
    category.append(i.split(','))
   
```

```python
for i in netflix['country']:
    countries.append(i.split(','))
    
```

```python
netflix.drop(['listed_in','country'],axis=1,inplace=True)
```

```python
netflix['category'],netflix['countries']=category,countries
```

```python
cast=[]
for i in netflix['cast']:
    cast.append(i.split(','))
netflix['cast']=cast    
```

```python
netflix.head()
```

```python
category_count=[]
for i in netflix.category:
    for j in i:
        category_count.append(j)

from nltk.probability import FreqDist
from heapq import nlargest

freq=FreqDist(category_count)
nlargest(10,freq,key=freq.get)
```

```python
freq
```

```python
countries_count=[]
for i in netflix['countries']:
    for j in i:
        countries_count.append(j)
freq2=FreqDist(countries_count)
freq2.pop('empty')
freq2
```

```python
cast_count=[]
for i in netflix['cast']:
    for j in i:
        cast_count.append(j)
freq3=FreqDist(cast_count)
freq3.pop('empty')
freq3
```
