# ****> Data Filling

```python
df.info()
```

```python
deaths=df['Total Deaths'].mean()
deaths
```

```python
df['Total Deaths']=df['Total Deaths'].fillna(int(deaths))
df['Total Deaths']
```

```python
deaths2=df['Deaths per 1M population'].mean()
deaths2
```

```python
df['Deaths per 1M population']=df['Deaths per 1M population'].fillna(int(deaths2))
df['Deaths per 1M population']
```

```python
tt=df['Total Tests'].mean()
tt
```

```python
df['Total Tests']=df['Total Tests'].fillna(int(tt))
df['Total Tests']
```

```python
tk=df['Tests per 1M population'].mean()
tk
```

```python
df['Tests per 1M population']=df['Tests per 1M population'].fillna(int(tk))
df['Tests per 1M population']
```
