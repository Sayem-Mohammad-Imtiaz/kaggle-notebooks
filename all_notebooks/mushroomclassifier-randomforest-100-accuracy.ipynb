```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# READING DATA

#### NOTE - I took na_values = '?' because there is an attribute which we will se furthur which determines '?' as a category so I think this might be a garbage value because it is affecting the dataset hence I'm dropping these values from dataset

```python
data = pd.read_csv('../input/mushroom-classification/mushrooms.csv',na_values = '?')
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#Here I'm dropping values equivalent to '?'
data = data.dropna()
```

# FEATURE PROCESSING
 
##### Since our dataset consists of categorical values in the form of characters so we need to convert them into numerical form hence we are doing feature processing.

```python
#We use Label Encoder and Ordinal Encoder for Feature Processing


from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
ordinal_enc = OrdinalEncoder()
label_enc = LabelEncoder()
```

```python
label_cat = data['class']
```

```python
label_class = label_enc.fit_transform(label_cat).reshape(-1,1)
```

```python
ord_cat = data.iloc[:,1:]
```

```python
ord_cat.columns
```

```python
ordinal_class = ordinal_enc.fit_transform(ord_cat)
```

```python
ordinal_class[0:5]
```

### Now after we are done with the feature processing, we are making a parallel dataset for applying our machine learning models on them
#### Our new dataset is a numerical dataset 

```python
NewData1 = pd.DataFrame(ordinal_class,columns = ord_cat.columns)
```

```python
NewData1.head()
```

```python
NewData2 = pd.DataFrame(label_class,columns = ["class"])
```

```python
NewData2.head()
```

```python
NewData = pd.concat([NewData1,NewData2],axis=1)
```

```python
NewData.head()
```

```python
NewData['class'] = NewData['class'].astype(float)
```

```python
NewData.head()
```

```python
NewData.columns
```

```python
x=NewData.iloc[:,:-1].values
```

```python
y = NewData['class'].values
```

# Splitting of data for training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=0)
```

# RANDOM FOREST CLASSIFIER

##### Firstly we are using a Random forest Classifier for classification of mushrooms in our mashrooms.csv dataset

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=200)
clf.fit(x_train,y_train)
```


### Accuracy Glitch

As we can see we are getting an accuracy of 100%.
In most of such situations there is a possibility of overfitting
So we will be checking the condition of overfitting furthur in our notebool

```python
clf.score(x_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(clf,x_train,y_train)
```

```python
cv_scores
```

# KNeighbors Classifier


```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 10)
```

```python
knn.fit(x_train,y_train)
```

## ACCURACY
#### Here we can see that using kNeighbors Classifier we are getting an accuracy of almost 99%

```python
knn.score(x_test,y_test)
```

```python
cv_scores2 = cross_val_score(knn,x_train,y_train)
```

```python
cv_scores2
```

```python
from sklearn.metrics import plot_confusion_matrix
```

## CONFUSION MATRIX FOR BETTER UNDERSTANDING OF OUR MODEL

```python
plot_confusion_matrix(knn, x_test, y_test, cmap="Greens", display_labels=np.asarray(data['class'].unique()))
plt.title("K-Nearest Neighbours (KNN) Confusion Matrix")
plt.xlabel("Predicted Mushroom")
plt.ylabel("Actual Mushroom")
```

```python
plot_confusion_matrix(clf, x_test, y_test, cmap="Greens", display_labels=np.asarray(data['class'].unique()))
plt.title("Random Forest Classifier")
plt.xlabel("Predicted Mushroom")
plt.ylabel("Actual Mushroom")
```
