## 1. Data Read and Data Review
### (Veri Okuma ve Veri İnceleme)

```python
data = pd.read_csv("/kaggle/input/forbes-billionaires-of-2021-20/forbes_billionaires.csv")
```

Veri Değişkenlerimiz : 
1. Name : Billionaire's Name (Milyarderin ismi)
2. NetWorth : (Wealth) Serveti
3. Country : Country (Ülkesi)
4. Source : Company name (Şirket(ler)inin Adı)
5. Ranks : Sequence (Milyarder sırası)
6. Age : Age (Yaşı)
7. Residence : Living Space (Yaşadığı şehir) 
8. Citizenship : Country of Citizenship (Vatandaşı olduğu ülke)
9. Status : Relationship status (İlişki Durumu)
10. Children : Number of children (Çocuk Sayısı)
11. Education : Education Status (Eğitim Durumu)
12. Self_made : is it his own gain (Emek Durumu)

## 5. Self Made - Wealth - Child Relationship
### (Kendi Emeği-Zenginlik-Çocuk İlişkisi)
