```python
def plotting_hist(var):
    plt.figure(figsize = (10,5))
    plt.hist(train_df[var], bins = 10)
    plt.xlabel('Columns')
    plt.ylabel("Frequency")
    plt.title("{} Distribution with Histogram".format(var))
    plt.show()
```

```python
numericVar = ["Fare", "Age","PassengerId","Sex","Pclass"]
for n in numericVar:
    plotting_hist(n)
```
