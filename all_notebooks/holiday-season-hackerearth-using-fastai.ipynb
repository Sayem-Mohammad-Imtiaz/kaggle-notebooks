```python
from fastai.vision.all import *
import torchvision.models as models
import numpy as np
import pandas as pd
import os
```

```python
train_df = pd.read_csv("../input/hackerearth-deep-learning-challenge-holidayseason/dataset/train.csv")
```

```python
train_df.head()
```

```python
dls = ImageDataLoaders.from_df(train_df,path="../input/hackerearth-deep-learning-challenge-holidayseason/dataset",
                               folder="train",
                               label_col="Class",valid_pct=0.2,
                               item_tfms= Resize(224),
                               batch_tfms=aug_transforms(),
                              bs=16)
```

```python
dls.vocab
```

```python
dls.show_batch()
```

```python
learn = cnn_learner(dls,models.resnet50,metrics=accuracy,pretrained=True)
```

```python
learn.fine_tune(8)
```

```python
test_path = "../input/hackerearth-deep-learning-challenge-holidayseason/dataset/test/"
test_images = os.listdir(test_path)
```

```python
test_path = "../input/hackerearth-deep-learning-challenge-holidayseason/dataset/test/"
test_image_ids = os.listdir(test_path)
test_images = [test_path+f"{img}" for img in test_image_ids]
preds = learn.get_preds(dl=dls.test_dl(test_images, shuffle=False, drop_last=False))
```

```python
preds = preds[0].cpu().numpy()
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
train_df["Class"] = lb.fit_transform(train_df.Class)
```

```python
pred = lb.inverse_transform((np.argmax(preds,axis=1)))
```

```python
submission = pd.DataFrame()
submission["Image"] = test_image_ids
submission['Class'] = pred
submission.to_csv("submission.csv", index=False)
```
