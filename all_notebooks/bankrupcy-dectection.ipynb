```python
#df2 = df1.interpolate(method='liear')
df2 = df1.interpolate(method='pad')
#['linear', 'time', 'index', 'values', 'nearest', 
#'zero', 'slinear', 'quadratic', 'cubic', 'barycentric', 'krogh',
#'spline', 'polynomial', 'from_derivatives', 'piecewise_polynomial', 'pchip', 'akima', 'cubicspline']
#nRow, nCol = df2.shape
#print(f'There are {nRow} rows and {nCol} columns')

print(df2.isnull().sum())
```

```python
# PREVIEW
pd.pivot_table(df2, index = ['class'],aggfunc = 'mean') # (mean, median, count)
```

```python
# UNDER SAMPLING
from imblearn.under_sampling import *

x_train, y_train = RandomUnderSampler(random_state=0).fit_sample(df2, df2['class'])
x_train = x_train.drop(['class'], axis =1)

pd.pivot_table(X_samp, index = y_samp, aggfunc = 'median')
```

```python
from sklearn.linear_model import LogisticRegression

from sklearn.metrics import classification_report

clf = LogisticRegression(random_state=77)
clf.fit(x_train, y_train)

```

```python
df_test = pd.read_csv('/kaggle/input/bankruptcy-detection/3year.csv', delimiter=',')
for i in range(1, 65):
    df_test[f'Attr{i}'] = pd.to_numeric(df_test[f'Attr{i}'], errors='coerce')

df_test = df_test.interpolate(method='pad')

x_test = df_test.drop(['class'], axis =1)
y_test = df_test['class']
```

```python
from sklearn.metrics import accuracy_score
y_predt = clf.predict(x_train)
print("logit train acc: {:.3f}".format(accuracy_score(y_predt, y_train)))
y_pred = clf.predict(x_test)
print("logit test acc: {:.3f}".format(accuracy_score(y_pred, y_test)))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
