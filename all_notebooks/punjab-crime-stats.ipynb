```python
#Code by Rashik Rahman  https://www.kaggle.com/rashikrahmanpritom/insights-on-disney-movies-gross

# Sort data by Styles in descending order 
CrimeType_desc = df.sort_values(by='CrimeType', ascending=False) 

# Display the top 10 Styles 
CrimeType_desc.head(10)
```

```python
#Code by Rashik Rahman https://www.kaggle.com/rashikrahmanpritom/insights-on-disney-movies-gross/notebook

# Extract year from release_date and store it in a new column
df['release_year'] = pd.DatetimeIndex(df['Year']).year

# Compute mean of adjusted gross per genre and per year
group = df.groupby(['CrimeType','release_year']).mean()

# Convert the GroupBy object to a DataFrame
CrimeType_yearly = group.reset_index()

# Inspect genre_yearly 
print(CrimeType_yearly.head())


# Plot the data  

g=sns.relplot(kind='line', x='release_year', y='CrimeCount', hue='CrimeType', data=CrimeType_yearly)
g.fig.set_size_inches(10,10)
plt.title('Crime Type trend')
```

#Dacoity

Dacoity is a term used for "banditry" in the Indian subcontinent. The spelling is the anglicised version of the Hindustani word daaku; "dacoit" /dəˈkɔɪt/ is a colloquial Indian English word with this meaning and it appears in the Glossary of Colloquial Anglo-Indian Words and Phrases (1903).

Banditry is criminal activity involving robbery by groups of armed bandits. The East India Company established the Thuggee and Dacoity Department in 1830, and the Thuggee and Dacoity Suppression Acts, 1836–1848 were enacted in British India under East India Company rule. Areas with ravines or forests, such as Chambal and Chilapata Forests, were once known for dacoits.

#Anti-Sikh riots 1984.

The 1984 anti-Sikh riots, also known as the 1984 Sikh Massacre, was a series of organised pogroms against Sikhs in India following the assassination of Indira Gandhi by her Sikh bodyguards. The ruling Indian National Congress had been in active complicity with the mob, as to the organisation of the riots. Government estimates project that about 2,800 Sikhs were killed in Delhi and 3,350 nationwide, whilst independent sources estimate the number of deaths at about 8,000–17,000.
