<h2 style='border:3px solid DodgerRed;background-color:#9d0aea;font-family:comic sans ms'><center>Detailed Understanding about Year column</center></h2>

```python
year = df['Year'].value_counts().reset_index()
```

```python
ff.create_table(year)
```

```python
px.pie(year,names="index",values="Year",color_discrete_sequence=px.colors.sequential.Bluyl_r)
```

<h2 style='border:3px solid DodgerRed;background-color:#9d0aea;font-family:comic sans ms'><center>Detailed Understanding about Genre column</center></h2>

```python
Genre = df["Genre"].value_counts().reset_index() 
Genre.columns = ["Genre","Value Counts"] 
px.pie(Genre,names="Genre",values="Value Counts",color_discrete_sequence=px.colors.sequential.Electric_r)
```

<h2 style='border:3px solid DodgerRed;background-color:#9d0aea;font-family:comic sans ms'><center>Detailed understanding about Lead studio</center></h2>

```python
lead = df["Lead Studio"].value_counts().reset_index() 
lead.columns = ["Lead Studio","Value Counts"] 
px.bar(lead,x="Lead Studio",y="Value Counts",color_discrete_sequence=px.colors.sequential.Greys_r)
```

<h2 style='border:3px solid DodgerRed;background-color:#9d0aea;font-family:comic sans ms'><center>Detailed understanding about Worldwide gross</center></h2>

```python
s = df["Worldwide Gross"].nlargest(10).index.tolist() 
f = df.iloc[s,:]
```

<h3 style = "font-family: Comic Sans MS;color:orange">Top 10 movie collections belonging to genre</h3>

```python
genre = f["Genre"].value_counts().reset_index() 
px.bar(genre,x="index",y="Genre",color_discrete_sequence=px.colors.sequential.Aggrnyl_r)
```

```python
hit_movies = f["Rotten Tomatoes %"].nlargest(3) 
flop_movies = f["Rotten Tomatoes %"].nsmallest(3)
```

```python
flop = df.iloc[flop_movies.index.to_list(),:]
```

<h3 style = "font-family: Comic Sans MS;color:orange">Top 3 flop movies based on Rotten Tomatoees score</h3>

```python
ff.create_table(flop)
```

<h3 style = "font-family: Comic Sans MS;color:green">Top 10 movies based on audience score</h3>
