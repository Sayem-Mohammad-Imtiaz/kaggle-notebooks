```python
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler


data = pd.read_csv('/kaggle/input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')
```

```python
# remove id 
data = data.drop('id', axis=1)
# remove na data by replacing with a 0
data = data.fillna(0)
```

```python
# take only the columns that needs to be encoded
data_to_encode = data[["gender", "ever_married",
                       "work_type", "Residence_type", "smoking_status"]]
# make it as list                       
df_col = list(data_to_encode.columns)
```

```python
# init label encoder
le = LabelEncoder()
# for every single column in df_col, encode the label
for i in range(len(df_col)):
    data[df_col[i]] = le.fit_transform(data[df_col[i]])
```

```python
# let's see correlation on the heatmap
hm = sns.heatmap(data.corr(),
                 cbar=True,  
                 square=True,  
                 yticklabels=data.columns,  
                 xticklabels=data.columns)  
```

```python
# prepare data 
X = data.drop('stroke', axis=1).values
Y = data['stroke'].values

X_train, X_test, Y_train, Y_test = train_test_split(
    X, Y, test_size=0.3, random_state=0)
```

```python
# standardize features
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
# let's build the model and train it
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

lr = LogisticRegression()
lr.fit(X_train, Y_train)
```

```python
# import evaluation data
from sklearn.metrics import accuracy_score 
from sklearn.metrics import log_loss 

Y_pred = lr.predict(X_test)
Y_pred_proba = lr.predict_proba(X_test)
print("all features LR")
print("ACCURACY: "+str(accuracy_score(Y_test, Y_pred)))
print("LOG LOSS: "+str(log_loss(Y_test, Y_pred_proba)))
```
