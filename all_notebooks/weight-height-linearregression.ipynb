# Statsmodels ile modelleme

```python
import statsmodels.api as sm
```

```python
X = data[['Height']]
X[0:5]
```

```python
X = sm.add_constant(X) # iki boyutlu yapıda olması için
X[0:5]
```

```python
y = data['Weight']
y[0:5]
```

```python
lm = sm.OLS(y,X)
model = lm.fit()
```

```python
model.summary()
```

```python
model.params
```

```python
model.mse_model
```

```python
print('Weight = ' + str('%.2f ' % model.params[0]) +  '+ Height * ' + str('%.2f' % model.params[1]))
```

```python
g = sns.regplot(data['Height'], data['Weight'], ci = None, scatter_kws = {'color':'r', 's':50})
g.set_title('Model Denklemi:Weight = -39.06 + Height * 61.27')
g.set_ylabel('Weight')
g.set_xlabel('Height')
import matplotlib.pyplot as plt
plt.xlim(1.4,1.9)
plt.ylim(45,80)
```

# sklearn ile linear_model

```python
from sklearn.linear_model import LinearRegression
```

```python
X = data[['Height']]
y = data['Weight']
reg = LinearRegression()
model = reg.fit(X,y)
```

```python
model.intercept_,model.coef_
```

```python
model.score(X,y)
```

```python
model.predict(X)[0:10]
```

# Artıklar

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import statsmodels.formula.api as smf
lm = smf.ols('Weight ~ Height', data)
model = lm.fit()
```

```python
mse = mean_squared_error(y, model.fittedvalues)
mse
```

```python
import numpy as np
rmse = np.sqrt(mse)
rmse
```
