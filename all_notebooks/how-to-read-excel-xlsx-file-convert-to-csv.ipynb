# Converting xlsx with multiple sheets into separate csv-s

This notebook provides the codes to -


[1. List the names of the sheets in the xlsx workbook](#p1)<br>
[2. Read each sheet in the workbook into pandas dataframes](#p2)<br>
[3. Export each sheet into a separate csv file.](#p3)<br>

```python
#install and import required packages 

!pip install openpyxl
import numpy as np 
import pandas as pd 
from openpyxl import load_workbook
```

<div id="p1">

### 1. List the names of the sheets in the xlsx workbook



```python
#Get name of each sheet from xlsx 
wb = load_workbook(filename = '../input/indian-states-sdg-index-202021/Data_SDG_India_Index_2020-21.xlsx')
sheetnames=wb.sheetnames

#Get number of sheets in the xlsx
n=len(sheetnames)
print(n)
```

### 2. Read each sheet in the workbook into pandas dataframes
<div id="p2">

```python
#Read data from xlsx sheets to dataframes
sdg=pd.read_excel('../input/indian-states-sdg-index-202021/Data_SDG_India_Index_2020-21.xlsx', 
              sheet_name=list(range(n)))

for i in range(1,n):
    sdg[i]  = pd.DataFrame(sdg[i].values[1:], columns=sdg[i].iloc[0])

sdg[0].head()
```

### 3. Export each sheet into a separate csv file.
<div id="p3">

```python

#convert sheets in the xlsx to a csv file
for i in range(n):
    sdg[i].to_csv('./'+sheetnames[i]+'.csv', index=False)
    
print("done!")
```
