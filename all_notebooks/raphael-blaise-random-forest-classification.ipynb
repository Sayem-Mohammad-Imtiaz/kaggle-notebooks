```python
dataset = pd.read_csv('../input/bank-customers-data/BankCustomerData.csv')
```

```python
# replacing unknown as null 
# The reason i am doing this is because i noticed some columns have unknown data which would not help
# with the acuuracy of my model and not make accurate predictions
dataset.replace(to_replace='unknown', value=np.NaN, inplace=True)
```

```python
# Dropping all the rows with NaN values for all the columns for 
#which the values for all the rows are Nan
dataset.dropna(axis=0, how ='all', inplace=True)
dataset.dropna(axis=1, inplace=True)
```

```python
# I picked those attributes one to prevent overfitting my model and two i tried to pick the ones
# which might have a co-relation to the subscription of term_deposit
dataset = dataset[['age','balance', 'month', 'duration', 'default',
                   'marital', 'loan', 'housing', 'day', 'term_deposit']]
```

```python
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# Encoding my categorical features in the matrix of features to numerical data
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encode', OneHotEncoder(),  [2, 4, 5, 6, 7])], remainder='passthrough' )
x = np.array(ct.fit_transform(x))
print(x)
```

```python
# Encoding my y values as they are in categorical data yes and no
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
```

```python
# Splitting my data into the training set and test set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
#Using the Random Forest Classifer i have to choose my n_estimators in this case number of trees and a 
#criterion to meausre the quality of the split
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators =10, criterion='entropy', random_state=0)
classifier.fit(x_train, y_train)
```

```python
# Predicting my results 
y_pred = classifier.predict(x_test)
print(y_pred)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
# Checking my correct and in-correct predictions using the confusion_matrix and then checking the accuracy of my model
# with the accuracy_score
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_pred, y_test)
print(cm)
print(accuracy_score(y_test, y_pred))
```
