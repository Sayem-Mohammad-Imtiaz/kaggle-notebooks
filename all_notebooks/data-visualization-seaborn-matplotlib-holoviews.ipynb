```python
import pandas as pd
import numpy as np
import hvplot.pandas

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
%matplotlib inline
```

```python
drinks_df = pd.read_csv("../input/drinks-by-country/drinksbycountry.csv")
drinks_df.head()
```

# Data exploration

# Data visualization

## Histograms

```python
drinks_df.hvplot.hist("total_litres_of_pure_alcohol", by="continent")
```

```python
# drinks_df.hvplot.scatter()
drinks_df.hvplot.hist("wine_servings", by="continent")
```

```python
drinks_df.hvplot.hist("beer_servings", by="continent")
```

```python
drinks_df.hvplot.hist("spirit_servings", by="continent")
```

```python
drinks_df.hvplot.hist("beer_servings", by="continent", subplots=True, width=250)
```

## Bar plots

```python
drinks_df.groupby('continent')['country'].count()
```

```python
drinks_df['continent'].value_counts().hvplot.bar()
```

## Kernel density estimate (kde)

```python
drinks_df.hvplot.kde(subplots=True, height=250, width=250)
```

## Scatter Plot

```python
drinks_df.hvplot.scatter(x='beer_servings', y='wine_servings', by='continent', 
                         legend='top', height=500, width=500)
```

```python
drinks_df.hvplot.area()
```

```python
sns.pairplot(drinks_df, hue='continent', height=3, aspect=1)
```

```python
sns.pairplot(drinks_df, hue='continent', height=3, diag_kind="hist");
```

```python
drinks_df.hist(edgecolor='black', linewidth=1.2, figsize=(12, 8));
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.violinplot(x='continent', y='beer_servings', data=drinks_df)

plt.subplot(2, 2, 2)
sns.violinplot(x='continent', y='spirit_servings', data=drinks_df)

plt.subplot(2, 2, 3)
sns.violinplot(x='continent', y='wine_servings', data=drinks_df)

plt.subplot(2, 2, 4)
sns.violinplot(x='continent', y='total_litres_of_pure_alcohol', data=drinks_df);
```

```python
drinks_df.boxplot(by='continent', figsize=(12, 8));
```

```python
pd.plotting.scatter_matrix(drinks_df, figsize=(12, 10));
```

```python
sns.pairplot(drinks_df)
```
