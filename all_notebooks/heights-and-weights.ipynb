### Import LinearRegression

```python
from sklearn.linear_model import LinearRegression
```

## Create the regression

#### Declare the dependent and independent variable

```python
x = data['Height']
y = data['Weight']
```

```python
x.shape
```

```python
y.shape
```

```python
x_matrix = x.values.reshape(15,1)  #creating 2D matrix
x_matrix.shape
```

#### Regression

```python
reg = LinearRegression()
```

```python
reg.fit(x_matrix,y)
```

#### R-Squared

```python
reg.score(x_matrix,y)
```

#### Coefficents

```python
reg.coef_
```

#### Intercept

```python
reg.intercept_
```

### Making Prediction

```python
new_data = pd.DataFrame(data=[2.3,3,4,5.3,4.3],columns=['Height'])
```

```python
new_data
```

```python
reg.predict(new_data)
```
