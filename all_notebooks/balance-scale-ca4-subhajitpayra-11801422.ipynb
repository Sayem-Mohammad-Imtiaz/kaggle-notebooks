```python
#Importing Libraries 
#basics and Visualization
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

#ML libraries
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split


#metrics
from statistics import mean
from sklearn.metrics import accuracy_score as score
from sklearn.metrics import explained_variance_score as evs

from sklearn.metrics import confusion_matrix
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score



#Ignore Warning 
import warnings as wrn
wrn.filterwarnings('ignore')
```

Reading the csv file

```python
df = pd.read_csv(r'../input/balance-scale.csv')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#Visualization after doing label encoding
df['Class'] = LabelEncoder().fit_transform(df['Class'].tolist())
#pairplot
sns.pairplot(data=df)

#Heatmap
num_feat = df.select_dtypes(include=np.number).columns
plt.figure(figsize= (15, 15))
sns.heatmap(df.corr())
```

Conclusion : Plots show that the other attributes are equally correlated to Class. 
- Hence, droping other columns except for 'Class' is not required

Seperating features as X and target value as y : 
- 'Class' in y
- 'L-Weight', 'R-Weight', 'L-Distance', 'R-Distance' in X

```python
#Dividing X and y
y = df[['Class']]
X = df.drop(['Class'], axis = 1)

print(y.head())
print(X.head())
```

Spliting the X and y in four DataFrame :
train_X, test_X, train_y and test_y using rain_test_split

```python
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 20)
```

Classification and prediction using following Classifiers :

- Extra Tree    Classifier
- Xgboost       Classifier
- Random Forest Classifier
- Decision Tree Classifier
- Support Vector Machine
- K-Nearest Neighbor

```python
#Classification and prediction
#ExtraTreeClassifier

clf = ExtraTreesClassifier(n_estimators=1000)
clf.fit(train_X, train_y)
pred1 = clf.predict(test_X)
print('Accuracy in percent = ',score(pred1, test_y)*100)
```

```python
confusion_matrix(test_y, pred1)
```

```python
print(classification_report(test_y, pred1))
```

```python
#Classification and prediction
#XGBoost

clf = XGBClassifier(learning_rate=0.5, n_jobs=-1, n_estimators=1000)
clf.fit(train_X, train_y)
pred2 = clf.predict(test_X)
print('Accuracy in percent = ',score(pred2, test_y)*100)
```

```python
confusion_matrix(test_y, pred2)
```

```python
print(classification_report(test_y, pred2))
```

```python
#Classification and prediction
#Random Forest

clf = RandomForestClassifier(n_estimators=1000, n_jobs=-1, )
clf.fit(train_X, train_y)
pred3 = clf.predict(test_X)
print('Accuracy in percent = ',score(pred3, test_y)*100)
```

```python
confusion_matrix(test_y, pred3)
```

```python
print(classification_report(test_y, pred3))
```

```python
#Classification and prediction
#DT

clf = DecisionTreeClassifier()
clf.fit(train_X, train_y)
pred4 = clf.predict(test_X)
print('Accuracy in percent = ',score(pred4, test_y)*100)
```

```python
confusion_matrix(test_y, pred4)
```

```python
print(classification_report(test_y, pred4))
```

```python
#Classification and prediction
#SVM

clf = SVC()
clf.fit(train_X, train_y)
pred5 = clf.predict(test_X)
print('Accuracy in percent = ',score(pred5, test_y)*100)
```

```python
confusion_matrix(test_y, pred5)
```

```python
print(classification_report(test_y, pred5))
```

```python
#Classification and prediction
#KNN

clf = KNeighborsClassifier(n_neighbors=9)
clf.fit(train_X, train_y)
pred6 = clf.predict(test_X)
print('Accuracy in percent = ',score(pred6, test_y)*100)
```

```python
confusion_matrix(test_y, pred6)
```

```python
print(classification_report(test_y, pred6))
```
