```python
data=pd.read_csv('/kaggle/input/breast-cancer-wisconsin-data/data.csv')

data.head()
```

```python
X=data.iloc[:,2:32].values
```

```python
X
```

```python
y=data.iloc[:,1].values

y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)

X_train
```

```python
X_test=scaler.fit_transform(X_test)

X_test
```

```python
np.sqrt(X_train.shape[0])
```

```python
k=21

from sklearn.neighbors import KNeighborsClassifier

knn=KNeighborsClassifier(n_jobs=k)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)

y_pred.shape
```

```python
y_test.shape
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test,y_pred)
```

```python
accuracy=[]

for i in range(1,50):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    accuracy.append(accuracy_score(y_test,knn.predict(X_test)))
    
```

```python
accuracy
```

```python
len(accuracy)
```

```python
plt.plot(range(1,50),accuracy)
```

```python
k=3

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

 **So we can see that when we select k=21 we get an  accuracy of 96 and when we select k=3 we get an accuracy of 98  hence value of k is to be selected as 3 for this knn classifier . **
