# Line plots

This makes a line plot. plt.show() displays your graph like print with text

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop)
plt.show()
```

plt.xlabel and plt.ylabel and plt.title make labels on the side

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')

plt.show()
```

The plt.ytick change the format of number on the left and below gives the label

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks([0, 2, 4, 6, 8, 10],           
           ['0', '2B', '4B', '6B', '8B', '10B'])

plt.show()
```

Another way to add ticks

```python
# Scatter plot
plt.plot(year, pop)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')

# Definition of tick_val and tick_lab
tick_val = [400, 21000, 50590]
tick_lab = ['15k', '10k', '100k']

# Adapt the ticks on the x-axis
plt.xticks(tick_val, tick_lab)

# After customizing, display the plot
plt.show()
```

pop = [1.0, 1.262, 1.650] + pop and year = [1800, 1850, 1900] + year adds more data to the graph

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.plot(year, pop)

# Add more data
year = [1800, 1850, 1900] + year
pop = [1.0, 1.262, 1.650] + pop

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')
plt.yticks([0, 2, 4, 6, 8, 10],           
           ['0', '2B', '4B', '6B', '8B', '10B'])

plt.show()
```

# Scatter matplotlib

If you replace .plot with .scatter it will make dots instead

```python
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]
plt.scatter(year, pop)
plt.show()
```

# Histograms

Replace .scatter with hist to make a histogram(a bar graph) you don't have to do it with bins

```python
year = [1950, 1970, 1990, 2010]
plt.hist(year, bins=10)
plt.show()
```

plt.clf()  cleans it up again so you can start afresh.

```python
# Build histogram with 5 bins
plt.hist(year, bins = 5)

# Show and clean up plot
plt.show()
plt.clf()

# Build histogram with 20 bins
plt.hist(year, bins = 20)

# Show and clean up again
plt.show()
plt.clf()
```
