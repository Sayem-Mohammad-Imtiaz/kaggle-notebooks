```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rcParams["figure.figsize"] = (18,9)
plt.rcParams['font.size']=18

tarr = [4.3, 4.1, 3.6, 0.0, 4.5, 4.4, 4.7, 4.2, 3.9, 4.6, 4.8, 3.2, 2.9, 3.1, 3.5, 3.7, 4.0, 5.0, 1.0, 4.9, 2.8, 3.3, 3.4, 2.3, 1.9, 2.7, 3.8, 3.0, 2.5, 2.2, 1.5, 2.0, 2.6, 2.4, 1.7, 1.6, 2.1, 1.8, 1.4, 1.3, 1.2, 1.1]
tarr.sort()

outs = [0]*len(tarr)

f = pd.read_csv("/kaggle/input/android-permission-dataset/Android_Permission.csv")

ratings = f['Rating']
clas = f['Class']

for i in range(1, len(ratings)):
    
    outs[tarr.index(ratings[i])] += int(clas[i])

plt.title("Class Rating and Amount")
plt.xlabel("Rating")
plt.ylabel("Class Amount")
plt.grid(linestyle="--", linewidth=1.3)
plt.plot(tarr, outs, marker="o", linewidth=1.618, color="grey")
plt.show()
```
