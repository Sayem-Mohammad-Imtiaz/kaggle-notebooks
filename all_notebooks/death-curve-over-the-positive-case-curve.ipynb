```python
corona_virus_brazil = pd.read_csv('../input/corona-virus-brazil/brazil_covid19.csv');
corona_virus_brazil_temp = pd.read_csv('../input/corona-virus-brazil/brazil_covid19.csv');
corona_virus_italy = pd.read_csv('../input/covid-19-italy-updated-regularly/national_data.csv');
corona_virus_usa = pd.read_csv('../input/covid19-in-usa/us_covid19_daily.csv');
print('Loaded')
```


The records of the Brazilian dataset are notified total cases added per day for each state. That is why it is necessary to collect the last notification of each stay to have the correct data.

Os registros do dataset brasileiro é notificado casos totais somados por dia para cada estado. Por isto é necessário coletar a última notificação de cada estadao para ter os dados corretos.


These first two graphs are a snapshot of the cases until one day before the date of the visualization. The subsequent visualizations are the evolution of the cases calculated by the sum of the cases until the day of the visualization.

Estes primeiros dois gráficos são um recorte dos casos até um dia anterior da data da visualização. As visualizações posteriores é a evolução dos casos calculado pela somatória dos casos até o dia da visualização.

**Total de casos por região (Gráfico 2) e Estados brasileiros (Gráfico 1)**

Total dos casos até a data anterior


**Total cases by region and state (Brazil)**


Total cases until the previous date

```python
plt.rcParams['figure.figsize'] = (20,7)
corona_virus_brazil_temp_state = corona_virus_brazil_temp.groupby('state').count()
corona_virus_brazil_temp_state = len(corona_virus_brazil_temp_state.index)
corona_virus_brazil_temp_total = corona_virus_brazil_temp.loc[ (len(corona_virus_brazil_temp) - corona_virus_brazil_temp_state) : len(corona_virus_brazil_temp) ]
corona_virus_brazil_temp_total = corona_virus_brazil_temp_total.reset_index()
corona_virus_brazil_temp_total.groupby('region').count()
corona_virus_brazil_temp_total_pie =  corona_virus_brazil_temp_total.loc[:,['region','cases']].groupby('region').sum()
corona_virus_brazil_temp_total_barh = corona_virus_brazil_temp_total.loc[:,['state','cases']].groupby('state').sum()
plt.subplot(121)
plt.barh(corona_virus_brazil_temp_total_barh.index, corona_virus_brazil_temp_total_barh['cases'])
plt.subplot(122)
plt.pie(corona_virus_brazil_temp_total_pie['cases'], autopct="%1.1f%%", labels=corona_virus_brazil_temp_total_pie.index)
plt.axis('equal') 
plt.show()
```

Remove columns not needed for analysis in the italian's dataset.

Remover colunas não necessárias para a análise no dataset italiano.

```python
corona_virus_italy_cases = pd.DataFrame(corona_virus_italy['data'],corona_virus_italy['totale_positivi']).rename(columns={'index': 'date', 'total_positive_cases': 'cases'})
corona_virus_italy_cases.tail()
print('Finished')
```

The data is grouped by date. In this way the records that took place in the same country in different states are added up.

Os dados são agrupados por data. Desta forma os registros que aconteceram no mesmo país em estados diferentes são somados.

```python
corona_virus_brazil_national_temp = corona_virus_brazil_temp.groupby('date').sum().reset_index()
corona_virus_italy_national_temp = corona_virus_italy.groupby('data').sum().reset_index()
corona_virus_usa_national_temp = corona_virus_usa.groupby('date').sum().reset_index()
corona_virus_usa[0:len(corona_virus_usa)]
corona_virus_brazil_national_temp.count()
corona_virus_italy_national_temp.count()
corona_virus_usa_national_temp.count()

print('Finished')
```

Comparação da evolução dos casos positivos de Coronavírus nos três países por intervalo de dias desde o primeiro caso até o momento. O eixo Y é a quantidade de casos positivos testados oficiais.

Comparison of numbers of positive cases of Coronavirus in three countries per day range from the first case to date. The Y-axis is a number of official tested positive cases.

```python
plt.figure(figsize=(15,10))

g = sns.lineplot(x=corona_virus_brazil_national_temp.index, y=corona_virus_brazil_national_temp['cases'], label="Brazil")
g = sns.lineplot(x=corona_virus_italy_national_temp.index, y=corona_virus_italy_national_temp['totale_positivi'], label="Italy")
g = sns.lineplot(x=corona_virus_usa_national_temp.index, y=corona_virus_usa_national_temp['positive'], label="USA")

plt.ylabel('Evolution of cases')
plt.xlabel('period( days )')
plt.title('Evolution of cases per day')
plt.show()

print('Finished')
```

Daily evolution of deaths in the three countries.

Evolução diária das mortes nos três países.

```python
plt.figure(figsize=(15,10))

sns.lineplot(x=corona_virus_brazil_national_temp.index, y=corona_virus_brazil_national_temp["deaths"], label='Brazil')
sns.lineplot(x=corona_virus_italy_national_temp.index, y=corona_virus_italy_national_temp["deceduti"], label='Italy')
sns.lineplot(x=corona_virus_usa_national_temp.index, y=corona_virus_usa_national_temp["death"], label='USA')
plt.xlabel("Daily Evolution index of deaths confirmed by the new Corona Virus")
plt.title('Evolution of deaths per day')
plt.show()

print('Finished')
```



Comparison between daily evolution of confirmed cases of deaths in Brazil, Italy and the United States. We can observe the progression of the death curve in relation to the number of positive cases and gain insights into the destructive potential of the increase in the death curve in relation to the increase in the positive case curve.

Comparação entre evolução diária de casos confirmados de mortes no Brasil, Itália e Estados Unidos. Podemos observar a progressão da curva de mortes em relação à quantidade de casos positivos e tirar insights do potencial destrutivo do aumento da curva de mortes em relação ao aumento da curva de casos positivos

```python
plt.figure(figsize=(15,10))

sns.lineplot(x='cases', y='deaths' , data=corona_virus_brazil_national_temp, label='Brazil' )
sns.lineplot(x='totale_positivi', y='deceduti' , data=corona_virus_italy_national_temp, label='Italy' )
sns.lineplot(x='positive', y='death' , data=corona_virus_usa_national_temp, label='USA' )
plt.ylabel('Number of deaths')
plt.xlabel('Positive cases evolution')
plt.title('Correlation between positive cases evolution and death numbers evolution')
plt.show()

print('Finished')
```

Conclusions - Conclusão

A correlation was observed between positive cases and number of deaths. The higher the line rises on the y-axis, the greater the evolution of deaths from positive cases.

Foi observado uma correlação entre casos positivos e número de mortes. Quanto maior a linha subir no eixo y maior a evolução de mortes por casos positivos.
