```python
shoe_normal_dist = df[df['prices.amountMin'] != df['prices.amountMax']]
nike_distr = shoe_normal_dist[shoe_normal_dist['brand']=='Nike']
nike_distr['avgPrice'] = (nike_distr['prices.amountMin'] + nike_distr['prices.amountMax'])/2
```

```python
x_min = nike_distr['avgPrice'].min()
x_max = nike_distr['avgPrice'].max()
np_arr =  nike_distr['avgPrice'].values
mean = np.mean(np_arr)
std = np.std(np_arr)
mean
std
```

```python
df.rename(columns={"prices.amountMin":"minPrice","prices.amountMax":"maxPrice"},inplace=True)
brand_shoe_df = df[["brand","minPrice","maxPrice"]]
brand_shoe_df['brand'] = brand_shoe_df['brand'].str.lower()
brand_shoe_df.head()
```

```python

average = df2.groupby('brand')['prices.amountMax'].mean().sort_values(ascending=False)

average
```

```python
og_data = df2[['brand', 'prices.amountMax']]
```

```python
# Looking into highest prices
og_data.sort_values(by='prices.amountMax', ascending=False)
```

```python
under500_data = og_data[og_data['prices.amountMax'] <= 500]
under500min5 = under500_data[under500_data.groupby('brand').brand.transform(len) > 4]
top15 = list(average.head(20)[1:].index)
data15 = under500min5.loc[under500min5['brand'].isin(top15)]
```

```python
# Now creating a violinplot for those same 15
plt.figure(figsize=(12,20))
sns.violinplot(y='brand', x='prices.amountMax', data=data15, order=top15).set_title('Top 15 Brands With Highest Price Averages')
plt.xlabel('Price Average (USD)')
plt.ylabel('Brand')
```

```python

asics_data = og_data.loc[og_data.brand == 'ASICS']

plt.figure()
sns.distplot(asics_data['prices.amountMax'], color='m').set_title('Asics Shoe Prices')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
```
