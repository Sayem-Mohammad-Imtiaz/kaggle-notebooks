* Regression is a type of supervised machine learning algorithm used to predict a continuous label. The goal is to produce a model that represents the 'best fit' to some observed data, according to an evaluation criterion.Here we use The Keras library to create the above stated model.
* We are going to make a Neural Network with 1 input layer and 2 hidden layers which could be altered easily.
* The End goal is to make an efficient Neural Network that is not overfitted and generalised for new sets of future data. 


# Importing libraries

```python
import pandas as pd
import numpy as np
```

```python
import keras
from keras.layers import Dense
from keras.models import Sequential
```

# Reading the Database

```python
df = pd.read_csv("../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df_cols = df.columns
```

# Creating Feature and Target Set

```python
predictor = df[df_cols[df_cols != "quality"]]
target = df[["quality"]]
```

# Normalizing The Data

```python
predictor_normal = ((predictor - predictor.mean())/predictor.std())
predictor_normal.head()
```

```python
n_cols = predictor_normal.shape[1]
```

# Creating The Regression Model (1 Input layer and 2 Hidden layers)

```python
def regression_model():
    model = Sequential()
    model.add(Dense(50,activation = "relu", input_shape =(n_cols,)))
    model.add(Dense(50,activation = "relu"))
    model.add(Dense(50,activation = "relu"))
    
    model.compile(optimizer = "adam", loss = "mean_squared_error")
    return model
```

# Training and Fitting

```python
model = regression_model()
model.fit(predictor_normal,target, validation_split = 0.3, epochs = 100, verbose = 1)
```

# Loss and evaluation

* After the 100 th epoch we see that the loss is **0.2960** and the validation loss is **0.4770** which is pretty decent.
