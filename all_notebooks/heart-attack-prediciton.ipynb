```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


sns.set_style("darkgrid")
plt.rcParams["figure.figsize"] = (12,5)
font = {"size"   : 11}

plt.rc('font', **font)

sns.countplot(x=heart_data["output"], data=heart_data);
```

```python
fig, axes = plt.subplots(2, 2, figsize=(25, 10))


sns.swarmplot(x="sex", y="chol", hue="output", data=heart_data, ax=axes[0][0])
axes[0][0].set_title("Chances of having a heart attack and by cholestorol degree and by sex");

sns.swarmplot(x="cp", y="chol", hue="output", data=heart_data, ax=axes[0][1])
axes[0][1].set_title("Chances of having a heart attack and by cholestorol degree, and by Chest pain");

sns.swarmplot(x="restecg", y="chol", hue="output", data=heart_data, ax=axes[1][0])
axes[1][0].set_title("Chances of having a heart attack and by cholestorol degree, and by resting electrocardiographic results");

sns.swarmplot(x="caa", y="chol", hue="output", data=heart_data, ax=axes[1][1])
axes[1][1].set_title("Chances of having a heart attack and by cholestorol degree, and by number of major vessels");
```

```python
fig, axes = plt.subplots(2, 2, figsize=(25, 10))


sns.swarmplot(x="sex", y="trtbps", hue="output", data=heart_data, ax=axes[0][0])
axes[0][0].set_title("Chances of having a heart attack and by resting blood pressure degree and by sex");

sns.swarmplot(x="cp", y="trtbps", hue="output", data=heart_data, ax=axes[0][1])
axes[0][1].set_title("Chances of having a heart attack and by resting blood pressure degree, and by Chest pain");

sns.swarmplot(x="restecg", y="trtbps", hue="output", data=heart_data, ax=axes[1][0])
axes[1][0].set_title("Chances of having a heart attack and by resting blood pressure degree, and by resting electrocardiographic results");

sns.swarmplot(x="caa", y="trtbps", hue="output", data=heart_data, ax=axes[1][1])
axes[1][1].set_title("Chances of having a heart attack and by cholestorol degree, and by number of major vessels");
```
