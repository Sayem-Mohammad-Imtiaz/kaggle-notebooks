# Introduction

<a id='Introduction'></a>

This presentation will examine SARS-CoV-2 (Covid-19) infection in Gibraltar. Gibraltar was selected for this presentation due to a very unique set of circumstances specific to the country.

<ul>
    <li>
        Gibraltar didn't have a death from covid-19 until November 11th, 2020.
    </li>
    <li>
        Gibraltar now has the highest death toll per capita in the world as of the time of this presentation. This death toll has happened in a short period of time.
    </li>
    <li>
        Gibraltar has been testing aggressively, with total tests several times that of the entire population.
    </li>
    <li>
        Gibraltar has been vaccinating aggressively since January 15th, 2021.
    </li>
    <li>
        Gibraltar has a median age that is very similar to the world median age.
    </li>
    <li>
        Gibraltar is a small country (6.8 square kilometres) with a low population (33,685) and high population density (~5000 per square kilometres) where transmission could occur very rapidly. It is possible that Gibraltar represents an accelerated timeline for what is happening everywhere else in the world.
    </li>
</ul>

This presentation will examine the possibility that Gibraltar is at or near a state of herd immunity for SARS-CoV-2.

Gibraltar has been testing aggressively during the entire period examined by this report. The regression function for total tests is virtually a straight line. This would suggest that Gibraltar was testing at full capacity during the period examined. This assertion is supported by the fact that pending tests increased significantly when infection rates began to spike.

<a href="#Daily_Tests" data-toc-modified-id="Daily_Tests">Link</a>

Infection rates appear to have accelerated starting around the middle of December 2020. This is where the peak of self-isolating cases can be seen. Testing facilities appear to become overwhelmed around the end of December 2020. All plots would suggest that the growth and decline of infections were exponential in nature. 

<a href="#Results_Pending" data-toc-modified-id="Results_Pending">Link</a>

Infection rates begin to go into decline around the third week of January. Average recovery time appears to be around twelve days. The worst period in terms of acceleration of deaths was the month of January.

The turning point appears to have been on or around January 18th, 2021. This date was used as the pivot point for regression models. On this date Gibraltar had a per capita death toll of 13 deaths per 10,000 in population.

The number of confirmed cases is unusual. First dose vaccinations didn't start until January 15th, 2021. Infections were already in decline at that point starting from around January 8th, 2021. Even those who were vaccinated starting on January 15th, 2021 wouldn't have received any benefit from it for some time. If the downturn in infections was the result of herd immunity starting to develop naturally in the population then the number of confirmed cases should be considerably higher considering how aggressively Gibraltar was testing. The alternative is that the mortality rate is much higher, and the results of this study do not support that conclusion. Another possible explanation is that the tests were returning a significant number of false negatives due to either the tests themselves or how the tests were being administered or processed. It is possible that a majority of the population had already acquired natural immunity prior to the introduction of widespread testing, but positive results even several weeks after infection is common. If this were the case then deaths probably would have began accelerating earlier. Another possibility is that confirmed cases were intentionally suppressed to accentuate the effectiveness of vaccinations and that most of those vaccinated had already acquire natural immunity.     

<a href="#Active_Cases" data-toc-modified-id="Active_Cases">Link</a>

It would appear that five days is approximately the period of time where an infected individual either recovers from infection or requires critical care. Average time spent in critical care appears to be around nine days for patients that recovered and five days for those that subsequently died. Emergency room visits peaked after critical care, suggesting that individuals requiring critical care will progress to the point quicker than individuals with milder symptoms.

There doesn't appear to be any meaningful difference between individuals that died from COVID as opposed to individuals that died where COVID was an exasperating factor to a pre-existing condition. Therefore, only total deaths will be considered for further analysis. It should be noted that the ratio of those that died directly from COVID versus individuals that died with COVID was about seven to one.

<a href="#Infection_Timeline" data-toc-modified-id="Infection_Timeline">Link</a>

The entire population of Gibraltar should have received a first dose by the middle of April, 2021 and second dose by the middle of May, 2021.

<a href="#First_Dose" data-toc-modified-id="First_Dose">Link</a>

Emergency room, COVID ward and critical care occupancy are predicted to reach zero in the second or third week of March. Active cases are predicted to come reach zero at the end of March or beginning of April. Self-Isolating cases are predicted to reach zero at the end of April or beginning of May. A lag of about a week can be observed between emergency room occupancy and either COVID ward or critical care occupancy. Gibraltar appears to be enforcing a fairly lengthy self-isolation period. Self-isolation is expected to zero almost two months after the zero date of emergency room occupancy and about a month after active cases are expected to reach zero.

<a href="#Predictions_Summary" data-toc-modified-id="Predictions_Summary">Link</a>

The greatest correlations exist between; daily cases, self-isolating cases, change in daily recovered cases, and active cases. The relationships between these metrics can be quantified with a predictability in low to mid 90% range.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pylab
import scipy.stats
import seaborn as sns
import statsmodels.api as sm
import warnings

from scipy import stats
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from statsmodels.stats.outliers_influence import OLSInfluence
from statsmodels.tsa.seasonal import seasonal_decompose

%matplotlib inline

warnings.filterwarnings('ignore')
sns.set()

class Gibraltar():
    
    file_data = ''
    file_data_datetime = ''
    infection_growth = ''
    infection_decline = ''
    infection_decline_second_dose = ''
    first_dose = ''
    
    date = 0
    day = 0
    tests = 0
    daily_tests = 0
    results_pending = 0
    results_received = 0
    confirmed_cases = 0
    daily_cases = 0
    active_cases = 0
    delta_active = 0
    recovered_cases = 0
    delta_recovered = 0
    self_isolation = 0
    delta_isolation = 0
    covid_ward = 0
    critical_care = 0
    emergency_room = 0
    deaths_from_covid = 0
    delta_deaths_from_covid = 0
    deaths_with_covid = 0
    delta_deaths_with_covid = 0
    total_deaths = 0
    delta_total_deaths = 0
    first_dose = 0
    delta_first_dose = 0
    second_dose = 0
    delta_second_dose = 0
    vaccines_administered = 0
    delta_vaccines_administered = 0
    
    population = 33684
    infection_pivot_date = '2021-01-18'
    first_dose_date = '2021-01-15'
    second_dose_date = '2021-02-01'
    
    data = np.array([])
    data_train = np.array([])
    data_train_scaled = np.array([])
    data_test = np.array([])    
    data_test_scaled = np.array([])
    
    prediction = np.array([])
    
    target_test = np.array([])    
    target_train = np.array([])    
    target = np.array([])
    
    predicted_zero_date = pd.DataFrame([])
    peak_date = pd.DataFrame([])
    peak_ratio = pd.DataFrame([])
    
    prediction_score = 0
    
    def get_max_trend(self, trend):
        
        max_index = 0
        max_trend = 0

        for trend_index, trend_value in trend.trend.items():
            if trend_value > max_trend:
                max_index = trend_index
                max_trend = trend_value

        print(max_index, max_trend)
        
        return max_index
    
    def get_max_ratio(self, ratio, ratio_description):
        
        max_index = 0
        max_value = 0

        for index, value in ratio.items():
    
            if value > max_value: 
        
                max_index = index
                max_value = value
    
        self.peak_ratio = self.peak_ratio.append({'Metric': ratio_description, 'Peak Date': max_index}, ignore_index = True)
        
        print(max_index, max_value)
        
        return self.file_data_datetime.loc[max_index][0]
    
    def predict_metric(self, index, metric, start_date, dose, metric_description):

        X = sm.add_constant(index)

        model = sm.OLS(metric, X)
        results = model.fit()
        print(results.summary())
        parameters = results.params
        print('y = ', parameters[1], 'x + ', parameters[0])

        sns.scatterplot(data = metric)
        plt.show()

        day = 0
        date = pd.to_datetime(start_date)
        daily_cases = 1

        while daily_cases > 0:

            day += 1
            date = date + pd.to_timedelta(1, unit = 'D')
            daily_cases = parameters[1] * day + parameters[0]

        self.predicted_zero_date = self.predicted_zero_date.append({'Dose': dose,
                                                                    'Metric': metric_description,
                                                                    'Zero Date': date},
                                                                    ignore_index = True)
        
        print('No data points after', date)
    
    def split_data(self):

        self.data_train, \
        self.data_test, \
        self.target_train, \
        self.target_test = train_test_split(self.data, \
                                            self.target, \
                                            test_size = 0.2)

        scaler = StandardScaler()
        scaler.fit(self.data_train)
        self.data_test_scaled = scaler.transform(self.data_test)
        self.data_train_scaled = scaler.transform(self.data_train)
    
    def run_algorithm(self, algorithm):

        regressor = algorithm
        regressor.fit(self.data_train_scaled, self.target_train)
        self.prediction = regressor.predict(self.data_test_scaled)
        self.prediction_score = regressor.score(self.data_test_scaled, self.target_test)
    
    def file_data_smoothed(self, days):
    
        file_data_smoothed = self.file_data_datetime.copy()
        
        file_data_smoothed['Daily_Tests'] = file_data_smoothed.Daily_Tests.rolling(window = days).mean()
        file_data_smoothed['Results_Pending'] = file_data_smoothed.Results_Pending.rolling(window = days).mean()
        file_data_smoothed['Results_Received'] = file_data_smoothed.Results_Received.rolling(window = days).mean()
        file_data_smoothed['Confirmed_Cases'] = file_data_smoothed.Confirmed_Cases.rolling(window = days).mean()
        file_data_smoothed['Daily_Cases'] = file_data_smoothed.Daily_Cases.rolling(window = days).mean()
        file_data_smoothed['Active_Cases'] = file_data_smoothed.Active_Cases.rolling(window = days).mean()
        file_data_smoothed['Delta_Active'] = file_data_smoothed.Delta_Active.rolling(window = days).mean()
        file_data_smoothed['Recovered_Cases'] = file_data_smoothed.Recovered_Cases.rolling(window = days).mean()
        file_data_smoothed['Delta_Recovered'] = file_data_smoothed.Delta_Recovered.rolling(window = days).mean()
        file_data_smoothed['Self_Isolation'] = file_data_smoothed.Self_Isolation.rolling(window = days).mean()
        file_data_smoothed['Delta_Isolation'] = file_data_smoothed.Delta_Isolation.rolling(window = days).mean()
        file_data_smoothed['COVID_Ward'] = file_data_smoothed.COVID_Ward.rolling(window = days).mean()
        file_data_smoothed['Critical_Care'] = file_data_smoothed.Critical_Care.rolling(window = days).mean()
        file_data_smoothed['Emergency_Room'] = file_data_smoothed.Emergency_Room.rolling(window = days).mean()
        file_data_smoothed['Deaths_From_COVID'] = file_data_smoothed.Deaths_From_COVID.rolling(window = days).mean()
        file_data_smoothed['Delta_Deaths_From_COVID'] = file_data_smoothed.Delta_Deaths_From_COVID.rolling(window = days).mean()
        file_data_smoothed['Deaths_With_COVID'] = file_data_smoothed.Deaths_With_COVID.rolling(window = days).mean()
        file_data_smoothed['Delta_Deaths_With_COVID'] = file_data_smoothed.Delta_Deaths_With_COVID.rolling(window = days).mean()
        file_data_smoothed['Total_Deaths'] = file_data_smoothed.Total_Deaths.rolling(window = days).mean()
        file_data_smoothed['Delta_Total_Deaths'] = file_data_smoothed.Delta_Total_Deaths.rolling(window = days).mean()
        file_data_smoothed['First_Dose'] = file_data_smoothed.First_Dose.rolling(window = days).mean()
        file_data_smoothed['Delta_First_Dose'] = file_data_smoothed.Delta_First_Dose.rolling(window = days).mean()
        file_data_smoothed['Second_Dose'] = file_data_smoothed.Second_Dose.rolling(window = days).mean()
        file_data_smoothed['Delta_Second_Dose'] = file_data_smoothed.Delta_Second_Dose.rolling(window = days).mean()
        file_data_smoothed['Vaccines_Administered'] = file_data_smoothed.Vaccines_Administered.rolling(window = days).mean()
        file_data_smoothed['Delta_Vaccines_Administered'] = file_data_smoothed.Delta_Vaccines_Administered.rolling(window = days).mean()
        
        return file_data_smoothed
    
    def regression(self, metric):
        
        response = sm.add_constant(self.day)
        predictor = metric

        model = sm.OLS(predictor, response)
        results = model.fit()
        print(results.summary())
        parameters = results.params
        print('response = ', parameters[1], 'predictor + ', parameters[0])
        
        influence = OLSInfluence(results)
        residuals = influence.resid_studentized_internal
        sns.boxplot(x = residuals)
        plt.show()
    
    def vaccination_plot(self, series1, series2):
    
        vaccinations = series1 / series2
        vaccinations_df = pd.DataFrame()
        vaccinations_df['Day'] = self.infection_decline.Day
        vaccinations_df['Vaccinations'] = vaccinations.rolling(window = 30).mean()
        sns.lineplot(x = vaccinations_df.Day, y = vaccinations_df.Vaccinations)
        plt.show()
    
    def ratio_plot(self, series1, series2, ratio_description):

        ratio_plot = series1 / series2

        ratio_plot_df = pd.DataFrame()
        ratio_plot_df['Day'] = self.file_data_datetime.Day
        ratio_plot_df['Ratio_Plot'] = ratio_plot.rolling(window = 30).mean()    
        peak_day = self.get_max_ratio(ratio_plot_df.Ratio_Plot, ratio_description)    
        sns.lineplot(x = ratio_plot_df.Day, y = ratio_plot_df.Ratio_Plot)
        plt.axvline(x = peak_day)
        plt.show()
    
    def load_file(self):
        
        self.file_data = pd.read_csv(self.file_name)
        self.date = self.file_data['Date']
        self.day = self.file_data['Day']
        self.tests = self.file_data['Tests']
        self.daily_tests = self.file_data['Daily_Tests']
        self.results_pending = self.file_data['Results_Pending']
        self.results_received = self.file_data['Results_Received']
        self.confirmed_cases = self.file_data['Confirmed_Cases']
        self.daily_cases = self.file_data['Daily_Cases']
        self.active_cases = self.file_data['Active_Cases']
        self.delta_active = self.file_data['Delta_Active']
        self.recovered_cases = self.file_data['Recovered_Cases']
        self.delta_recovered = self.file_data['Delta_Recovered']
        self.self_isolation = self.file_data['Self_Isolation']
        self.delta_isolation = self.file_data['Delta_Isolation']
        self.covid_ward = self.file_data['COVID_Ward']
        self.critical_care = self.file_data['Critical_Care']
        self.emergency_room = self.file_data['Emergency_Room']
        self.deaths_from_covid = self.file_data['Deaths_From_COVID']
        self.delta_deaths_from_covid = self.file_data['Delta_Deaths_From_COVID']
        self.deaths_with_covid = self.file_data['Deaths_With_COVID']
        self.delta_deaths_with_covid = self.file_data['Delta_Deaths_With_COVID']
        self.total_deaths = self.file_data['Total_Deaths']
        self.delta_total_deaths = self.file_data['Delta_Total_Deaths']
        self.first_dose = self.file_data['First_Dose']
        self.delta_first_dose = self.file_data['Delta_First_Dose']
        self.second_dose = self.file_data['Second_Dose']
        self.delta_second_dose = self.file_data['Delta_Second_Dose']
        self.vaccines_administered = self.file_data['Vaccines_Administered']
        self.delta_vaccines_administered = self.file_data['Delta_Vaccines_Administered']
        
        self.file_data_datetime = self.file_data.copy()
        self.file_data_datetime['Date'] = pd.to_datetime(self.file_data_datetime['Date'])
        self.file_data_datetime = self.file_data_datetime.set_index('Date').asfreq('d')
        
        self.infection_growth = gibraltar.file_data.loc[:109].copy()
        self.infection_growth['Day'] = range(1, (len(self.infection_growth) + 1))
        
        self.infection_decline = gibraltar.file_data.loc[109:].copy()
        self.infection_decline['Day'] = range(1, (len(self.infection_decline) + 1))
        self.infection_decline['Active_Cases_Sqrt'] = np.sqrt(self.infection_decline['Active_Cases'])
        self.infection_decline['Self_Isolation_Sqrt'] = np.sqrt(self.infection_decline['Self_Isolation'])
        self.infection_decline['COVID_Ward_Sqrt'] = np.sqrt(self.infection_decline['COVID_Ward'])
        self.infection_decline['Critical_Care_Sqrt'] = np.sqrt(self.infection_decline['Critical_Care'])
        self.infection_decline['Emergency_Room_Sqrt'] = np.sqrt(self.infection_decline['Emergency_Room'])
        
        self.infection_decline_second_dose = gibraltar.file_data.loc[123:].copy()
        self.infection_decline_second_dose['Day'] = range(1, (len(self.infection_decline_second_dose) + 1))
        self.infection_decline_second_dose['Active_Cases_Sqrt'] = np.sqrt(self.infection_decline_second_dose['Active_Cases'])
        self.infection_decline_second_dose['Self_Isolation_Sqrt'] = np.sqrt(self.infection_decline_second_dose['Self_Isolation'])
        self.infection_decline_second_dose['COVID_Ward_Sqrt'] = np.sqrt(self.infection_decline_second_dose['COVID_Ward'])
        self.infection_decline_second_dose['Critical_Care_Sqrt'] = np.sqrt(self.infection_decline_second_dose['Critical_Care'])
        self.infection_decline_second_dose['Emergency_Room_Sqrt'] = np.sqrt(self.infection_decline_second_dose['Emergency_Room'])

        self.first_dose = gibraltar.file_data.loc[106:].copy()
        self.first_dose['Day'] = range(1, (len(self.first_dose) + 1))
        
        self.file_data.info()  
        
    def __init__(self):
        
        pass
    
gibraltar = Gibraltar()
gibraltar.file_name = '../input/sarscov2gibraltar/Covid_Data.csv'
gibraltar.load_file()
gibraltar.file_data.isnull().sum()
```
