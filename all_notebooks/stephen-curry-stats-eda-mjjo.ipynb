### Analysis of Season stats

```python
def highlight_max(data, color=main_colors[1]):
    attr = 'color: {}'.format(color)
    if data.ndim == 1:  # Series from .apply(axis=0) or axis=1
        is_max = data == data.max()
        return [attr if v else '' for v in is_max]
    else:  # from .apply(axis=None)
        is_max = data == data.max().max()
        return pd.DataFrame(np.where(is_max, attr, ''),
                            index=data.index, columns=data.columns)

regular_season_x = regular_season.drop(['Season_div', 'Date', 'OPP', 'Result', 'T Score', 'O Score', 'FG', '3PT', 'FT'], axis=1)
regular_stats2 = pd.DataFrame(columns=list(regular_season_x.columns))
seasons = regular_season['Season_year'].unique()

for s in seasons:
    regular_stats2.loc[s] = regular_season_x.loc[regular_season_x['Season_year'] == s, :].sum().values.reshape(-1)

games = []
for s in seasons:
    games.append(
    regular_season_x.loc[regular_season_x['Season_year'] == s, 'MIN'].count())

regular_stats2 = regular_stats2.drop(['Season_year', 'FG%', '3P%', 'FT%'], axis=1)
regular_stats2['Game'] = games
regular_stats2 = regular_stats2[['Game', 'MIN', 'FGM', 'FGA', '3PTM', '3PTA', 'FTM', 'FTA', 'REB', 'AST', 'BLK', 'STL', 'PF', 'TO', 'PTS']]
regular_stats2.style.format('{:.1f}').apply(highlight_max)
```

```python
regular_stats2.loc['Total'] = regular_stats2.sum().values
regular_stats2.loc[['Total']]
```

```python
regular_stats = pd.DataFrame(columns=list(regular_season_x.columns))
regular_stats = regular_stats.drop('Season_year', axis=1)
seasons = regular_season_x['Season_year'].unique()

regular_stats['FG%'] = regular_stats2['FGM'] / regular_stats2['FGA'] * 100
regular_stats['3P%'] = regular_stats2['3PTM'] / regular_stats2['3PTA'] * 100
regular_stats['FT%'] = regular_stats2['FTM'] / regular_stats2['FTA'] * 100
regular_stats = regular_stats.drop('Total', axis=0)

for s in seasons:
    regular_stats.loc[s] = regular_season_x.loc[regular_season_x['Season_year'] == s, :].mean().round(2).values.reshape(-1)

regular_stats.style.format('{:.1f}').apply(highlight_max)
```

### What I found out


* Curry's average playing minute is highest in 2012-2013 season. Generally his playing time is in 33-36 minutes. After Steve Kerr's appointed to director of GSW, his playing minutes is almost under 35 minutes

* Curry's average assists, turnovers are highest in 2013-2014 season, which is thought to be a result of his high proportion of ball possession due to the PG(Point Guard) position's characteristics. Also, after 2013-2014 season, his number of average assists decrease about two, That's beause his teammate Draymond Green was playing as a playmaker with him, and Curry transformed into a more scoring-focused play style.

* Curry's average FG%(field goals percentage), 3PT%(3 points field goals percentage) and steals are highest 2015-2016 season which was Curry's unanimous MVP. Also the almost season stats (attempted field goals, made field goals, attempted 3 points field goals, made 3 points field goals, attempted free throws, made free throws, rebounds, steals, points) are the maximum in that season.

* Most other average records are the highest in 2020-2021 season. Especially, Curry is making records about offense stats(average attempted field goals, average made field goals, average attempted 3 points field goals, average made 3 points field goals, average attempted free throws, average made free throws, average points). The absence of Klay Thompson who is GSW's second option player is the biggest  reason for that records

```python
def highlight_min(data, color=main_colors[0]):
    attr = 'color: {}'.format(color)
    if data.ndim == 1:  # Series from .apply(axis=0) or axis=1
        is_min = data == data.min()
        return [attr if v else '' for v in is_min]
    else:  # from .apply(axis=None)
        is_min = data == data.min().min()
        return pd.DataFrame(np.where(is_min, attr, ''),
                            index=data.index, columns=data.columns)

regular_team = regular_season.groupby('OPP').mean()
regular_team['Games'] = regular_season[['Result', 'OPP']].groupby('OPP').count().values
regular_team['W'] = regular_season.loc[regular_season['Result']=='W', ['Result', 'OPP']].groupby('OPP').count().values
regular_team['L'] = regular_season.loc[regular_season['Result']=='L', ['Result', 'OPP']].groupby('OPP').count().values
regular_team['W/L'] = regular_team['W'] - regular_team['L']
regular_team['T/O Score'] = regular_team['T Score'] - regular_team['O Score']

team_stats = ['Games', 'W', 'L', 'W/L', 'T Score', 'O Score', 'T/O Score']
```
