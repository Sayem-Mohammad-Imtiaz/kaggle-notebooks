```python
df_train['Gender'] = df_train['Gender'].map({'Female' : 0, 'Male' : 1})
df_train['Customer Type'] = df_train['Customer Type'].map({'Loyal Customer' : 0, 'disloyal Customer' : 1})
df_train['Type of Travel'] = df_train['Type of Travel'].map({'Business travel' : 0, 'Personal travel' : 1})
df_train['Class'] = df_train['Class'].map({'Eco' : 0, 'Business' : 1})
df_train['satisfaction'] = df_train['satisfaction'].map({'neutral or dissatisfied' : 0, 'satisfied' : 1})

df_test['Gender'] = df_test['Gender'].map({'Female' : 0, 'Male' : 1})
df_test['Customer Type'] = df_test['Customer Type'].map({'Loyal Customer' : 0, 'disloyal Customer' : 1})
df_test['Type of Travel'] = df_test['Type of Travel'].map({'Business travel' : 0, 'Personal travel' : 1})
df_test['Class'] = df_test['Class'].map({'Eco' : 0, 'Business' : 1})
df_test['satisfaction'] = df_test['satisfaction'].map({'neutral or dissatisfied' : 0, 'satisfied' : 1})
```

```python
from sklearn.model_selection import train_test_split
import xgboost as xgb
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
import statistics

target = df_train['satisfaction']
test_data = df_test[['id', 'Gender', 'Customer Type', 'Age', 'Type of Travel', 'Class', 'Flight Distance', 'Inflight wifi service', 'Departure/Arrival time convenient', 'Ease of Online booking', 'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort', 'Inflight entertainment', 'On-board service', 'Leg room service', 'Baggage handling', 'Checkin service', 'Inflight service', 'Cleanliness', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']]
train_data = df_train[['id', 'Gender', 'Customer Type', 'Age', 'Type of Travel', 'Class', 'Flight Distance', 'Inflight wifi service', 'Departure/Arrival time convenient', 'Ease of Online booking', 'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort', 'Inflight entertainment', 'On-board service', 'Leg room service', 'Baggage handling', 'Checkin service', 'Inflight service', 'Cleanliness', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']]

results, predictions = [], []

X_train, X_test, y_train, y_test = train_test_split(train_data, target, test_size=0.3, random_state=42)

model = xgb.XGBClassifier(n_estimators=5000,max_depth=5,reg_lambda=0.8, random_state=42)

model.fit(X_train, y_train, early_stopping_rounds=300, eval_set=[(X_test, y_test)], verbose=False)
y_prediction = model.predict(X_test)

results.append(accuracy_score(y_test,y_prediction))

print(results)
```
