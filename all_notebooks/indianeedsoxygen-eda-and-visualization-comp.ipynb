```python
#importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#reading/loading dataset
df= pd.read_csv("../input/indianeedsoxygen-tweets/IndiaWantsOxygen.csv")
df= df[pd.notnull(df["user_name"])]
```

```python
def clean_text(inp,words_):
    out = inp
    for i in words_:
        out = out.replace(i," ")
    return out
```

```python
words_= "".join(df["text"])
words = clean_text(words_,["https://",".co","\n","  ","IndiaNeedOxygen"])
```

```python
#wordcloud visualisation of texts within the tweets
from wordcloud import WordCloud, STOPWORDS
plt.figure(figsize= (20,20))
final = WordCloud(width = 2000, height = 800, background_color ="black",min_font_size = 10).generate(words)
plt.imshow(final)
plt.axis("off") 
plt.show()
```
