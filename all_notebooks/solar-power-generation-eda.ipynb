```python
#Merging dataframes
df_all_gen_data = pd.concat([df_p1_gen_data, df_p2_gen_data])
df_all_weather_data = pd.concat([df_p1_weather_data, df_p2_weather_data])

#What is the mean value of daily yield?
df_all_gen_data.groupby("PLANT_ID")["DAILY_YIELD"].mean()
```

```python
#Total of irradition group by plant.
df_all_weather_data.groupby("PLANT_ID").sum()
```

```python
#What is the total irradiation per day?
df_p1_weather_data.groupby(df_p1_weather_data["DATE_TIME"].str[0:11])["IRRADIATION"].sum()
```

```python
#What is the max ambient and module temperature?
df_all_weather_data.groupby("PLANT_ID")[["AMBIENT_TEMPERATURE", "MODULE_TEMPERATURE"]].max()
```

```python
#How many inverters are there for each plant?
df_all_gen_data.groupby(by="PLANT_ID")["SOURCE_KEY"].nunique()
```

```python
#What is the maximum/minimum amount of DC/AC Power generated in a time interval/day?
df_all_gen_data.groupby(df_all_gen_data["DATE_TIME"].str[0:11])[["DC_POWER", "AC_POWER"]].max()
```

```python
df_all_gen_data.groupby(df_all_gen_data["DATE_TIME"].str[0:11])[["DC_POWER", "AC_POWER"]].min()
```

```python
#Which inverter (source_key) has produced maximum DC/AC power?
df_all_gen_data.groupby("SOURCE_KEY").sum().sort_values("DC_POWER", ascending=False)
```

```python
#Rank the inverters based on the DC power they produce
df_all_gen_data.groupby("SOURCE_KEY")["DC_POWER"].sum().rank(ascending=False).sort_values()
```

```python
#Rank the inverters based on the AC power they produce
df_all_gen_data.groupby("SOURCE_KEY")["AC_POWER"].sum().rank(ascending=False).sort_values()
```
