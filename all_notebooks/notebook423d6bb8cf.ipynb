```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Now lets look at what data we are dealing with in order to see what I want to plot 

```python
menu = pd.read_csv('../input/Netflix Shows.csv',encoding='cp1252')
menu.info()
```

Now it seems from the information that rating should be good to create a scatter plot!
Here is the code to create it:

```python
measuring=['user rating score']
for m in measuring:
    plot=sns.swarmplot(x='rating',y=m,data=menu)
    plt.setp(plot.get_xticklabels(),rotation=35)
    plt.title('TV ratings for dataset')
    plt.show()
```
