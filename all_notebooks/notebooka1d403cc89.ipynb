### Obtén una lista con los atributos cuantitativos y otra lista con los atributos cualitativos.

```python
df = pd.read_csv('/kaggle/input/the-human-freedom-index/hfi_cc_2020.csv')
df.drop('Unnamed: 0', axis=1, inplace = True)
print(df.info())
df.head()
```

Atributos cuantitativos: 109 (float64, int64)
Atributos cualitativos: 3 (object)

```python
series_objetos = (df.dtypes == 'object').values
series_no_obj = [not i for i in series_objetos]
```

```python
atributos_cual = df.iloc[0,series_objetos].index
atributos_cual
```

```python
atributos_cuant = df.iloc[0, series_no_obj].index
atributos_cuant
```

### Define el rango de cada uno de los atributos (cuantitativos).

```python
rango_atrb = dict()
for i in atributos_cuant:
    rango_atrb[i] = max(df[i])-min(df[i])
print(rango_atrb)
```

### Obtén la moda o promedio dependiendo el caso para cada atributo.

```python
# Moda: Cualitativos
from scipy import stats
modas = dict()
for i in atributos_cual:
    modas[i] = stats.mode(df[i])[0][0]
modas
```

```python
# Promedio: Cuantitativos
promedios = dict()
for i in atributos_cuant:
    promedios[i] = np.mean(df[i])
print(promedios)
```

### Obtén el rango, varianza, desviación estandar para los atributos cuantitativos.

```python
# Rango
print(rango_atrb)
```

```python
# Varianza
var_cuant = dict()
for i in atributos_cuant:
    var_cuant[i] = np.var(df[i])
print(var_cuant)
```

```python
# Desviación estándar
std_cuant = dict()
for i in atributos_cuant:
    std_cuant[i] = np.std(df[i])
print(std_cuant)
```

### Calcula el porcentaje de datos dentro de 1 desviación standard y dentro de 2 desviaciones standard

```python
# Medias aritméticas
mean_cuant = dict()
for i in atributos_cuant:
    mean_cuant[i] = np.mean(df[i])
```

```python
# Función para *x* desviaciones estándar
def rango_desv_st(n):
    rango_std = dict()
    dentro_rango = dict()
    for i in atributos_cuant:
        rango_std[i] = (np.mean(df[i])-std_cuant[i]*n, np.mean(df[i])+std_cuant[i]*n)
        dentro_rango[i] = len([None for valor in df[i] if (valor>=rango_std[i][0] and valor<=rango_std[i][1])])/len(df)*100
        dentro_rango[i] = str(round(dentro_rango[i], 2)) + '%'
    return(dentro_rango)
```

```python
# 1 desviación estándar
print(rango_desv_st(1))
```

```python
# 2 desviación estándar
print(rango_desv_st(2))
```
