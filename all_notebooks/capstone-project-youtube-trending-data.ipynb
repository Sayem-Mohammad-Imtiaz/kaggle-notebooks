## which year was data collected

```python
cdf = df['trending_date'].apply(lambda x : '20'+x[:2]).value_counts().to_frame().reset_index().rename(columns = {"index":"year","trending_date":"no_of_videos"})
    
```

```python
print(cdf)
```

```python
#bar chart

plt.bar(cdf['year'],cdf['no_of_videos'])
```

```python
#what is the percentage of the videos released in the particular year

df['trending_date'].apply(lambda x : '20'+x[:2]).value_counts(normalize = True)
```

```python
#descriptiion

df.describe()
```

```python
df.hist('views')
```

```python
#number of videos less than 1million views

df[df['views']<2e6]['views'].count()/df['views'].count()*100
```

```python
df.hist('likes')
```

```python
#percentage of videos with >50k likes

df[df['likes']>50000]['likes'].count()/df['likes'].count()*100
```

```python
#percentage of videos with >10k dislikes

df[df['dislikes']>10000]['dislikes'].count()/df['dislikes'].count()*100
```

```python
#percentage of videos with >10k comment_count

df[df['comment_count']>10000]['comment_count'].count()/df['comment_count'].count()*100
```

```python
#discription of non numerical columns

df.describe(include = 'O')
```

```python
#how do we calculate the title length

df['title_length'] = df['title'].apply(lambda x : len(x))
```
