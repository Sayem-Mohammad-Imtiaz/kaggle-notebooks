```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go

from IPython.display import display
```

```python
date = '062921'
```

## BSE All Stocks

```python
all_df_bse = pd.read_csv(f'/kaggle/input/nse-bse-stocks-daily-change/{date}_bse.csv')

fig = px.scatter(all_df_bse, x="Company Name", y="% Change", color="Category")

fig.update_layout(title='BSE % Change', showlegend=False)

fig.update_layout(xaxis=dict(rangeslider=dict(visible=True)))

fig.show()
```

```python
# Add Color Attribute based on their '% Change'

def add_Color(df):
    attr = '% Change'
    color = 'Color' 
    df[color] = 0

    for i in range(len(df)):
        if df.loc[i, attr] > -5.0: df.loc[i, color] = 1

        if df.loc[i, attr] > -2.5: df.loc[i, color] = 2

        if df.loc[i, attr] > 0.0: df.loc[i, color] = 3

        if df.loc[i, attr] > 2.5: df.loc[i, color] = 4

        if df.loc[i, attr] > 5.0: df.loc[i, color] = 5
        
    return df
```

## BSE Stocks based on Categories

```python
chg_df_bse = all_df_bse[['% Change','Color','Category']].groupby(by=['Category'], as_index=False).mean()

chg_df_bse = add_Color(chg_df_bse)

fig = px.scatter(chg_df_bse, x="Category", y="% Change",  color="Color")

fig.update_layout(title='BSE % Change',height=800)

fig.update_layout(xaxis=dict(rangeslider=dict(visible=True)))

fig.show()
```
