```python
#Assiduidade dos vereadores - painel eletronico
#Dados da CÃ¢mara de Feira de Santana - BA
#https://www.kaggle.com/dadosabertosdefeira/assiduidade-dos-vereadores-painel-eletronico/code
# read json files and generate a dataframe that is saved to csv

import json
import pandas as pd
import os

def getAttendanceDataframe(jsondata):
    #iterate top level
    top={}
    for key in jsondata:
        if not isinstance(jsondata[key],dict):
            #print(key,":",jsondata[key])
            top[key]=jsondata[key]

    #iterate attendance
    data = []
    for p in jsondata['attendance']['attending']:
        #print(p)
        newp = top.copy()
        newp.update(p)
        newp['attendance']='attending'
        newp['text']=''
        data.append(newp)

    for p in jsondata['attendance']['absent']:
        #print(p)
        newp = top.copy()
        newp.update(p)
        newp['attendance']='absent'
        newp['text']=''
        data.append(newp)

    for p in jsondata['attendance']['justified']:
        #print(p)
        newp = top.copy()
        newp.update(p)
        newp['attendance']='justified'
        data.append(newp)

    df = pd.DataFrame(data)
    #display(df)

    return df


allattendance = pd.DataFrame()

for dirname, _, filenames in os.walk('/kaggle/input'):
    jsonfiles = [ fi for fi in filenames if fi.endswith(".json") ]
    for filename in jsonfiles:
        filename=os.path.join(dirname, filename)
        print(filename)
        f = open(filename, 'r')
        file_contents = json.load(f)
        f.close()
        df = getAttendanceDataframe(file_contents)
        print("generated (lines,columns): ",df.shape)
        allattendance = allattendance.append(df)

allattendance.to_csv("Assiduidade_dos_vereadores_dadosabertosdefeira.csv")

       
```
