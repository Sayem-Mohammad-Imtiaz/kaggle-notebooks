```python
df = pd.read_csv(pathToFile, sep=',')

data = df.to_numpy()

x = data[:, :-1]     # take all but except last column
y = data[:,-1]       # only last column => target



```

```python
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.svm import SVR
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

numerical_columns = [5, 6]
numeric_transformer = Pipeline(
    steps=[
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ]
)
categorical_columns = [0, 1, 2, 3, 4]
categorical_transformer = OneHotEncoder(handle_unknown='ignore')
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numerical_columns),
        ('cat', categorical_transformer, categorical_columns)
    ]
)

clf = Pipeline(
    steps=[
        ('preprocessor', preprocessor),
        ('classifier', SVR(degree=6, kernel='linear', verbose=2))
    ]
)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)

clf.fit(x_train, y_train)
score = clf.score(x_test, y_test)
pred = clf.predict(x_test)

print(f"Score poly: {score}")
print(f"Pred: {pred}")
```
