```python
links_df = pd.read_csv("/kaggle/input/movielens/links.csv")
movies_df = pd.read_csv("/kaggle/input/movielens/movies.csv")
ratings_df = pd.read_csv("/kaggle/input/movielens/ratings.csv")
tags_df = pd.read_csv("/kaggle/input/movielens/tags.csv")
```

```python
# Split year from the title column in the movies_df DataFrame
movies_df["year"] = movies_df["title"].str.extract('.*\((.*)\).*', expand=True)
movies_df.head()
```

```python
# Merge movies with ratings DataFrame
ave_ratings = ratings_df.groupby("movieId", as_index=False).mean()
merged_df = movies_df.merge(ave_ratings, on="movieId", how="inner")
merged_df = merged_df[["year","title", "rating", "genres"]]
merged_df.head()
```

```python
# Create mask for genres
genre_category_animation = "Animation"
genre_category_sci_fi = "Sci"
genre_category_horror = "Horror"


# Merge movies with tags and ratings DataFrame
# average_rating = ratings_df[["movieId", "rating"]].groupby("movieId",as_index=False)
ave_ratings = ratings_df.groupby("movieId", as_index=False).mean()
merged_df = movies_df.merge(ave_ratings, on="movieId", how="inner")
merged_df = merged_df[["year","title", "rating", "genres"]]

merged_df.head()
```

```python
mask1 = merged_df["genres"].str.contains(genre_category_animation)
mask2 = merged_df["genres"].str.contains(genre_category_sci_fi)
mask3 = merged_df["genres"].str.contains(genre_category_horror)


animation_movies = merged_df[mask1]
sci_fi_movies = merged_df[mask2]
horror_movies = merged_df[mask3]
```

```python
animation_movies.head()
```

```python
ave_animation_movies_ratings_per_year = animation_movies.groupby("year").mean()
ave_sci_fi_movies_ratings_per_year = sci_fi_movies.groupby("year").mean()
ave_horror_movies_ratings_per_year = horror_movies.groupby("year").mean()
```

```python
import matplotlib.pyplot as plt
import matplotlib.style

# Plot final visualization
ave_merged_df = pd.concat([ave_animation_movies_ratings_per_year,
                           ave_sci_fi_movies_ratings_per_year,
                           ave_horror_movies_ratings_per_year], axis=1, join="inner")

ave_merged_df.columns = ["Animation", "Sci-fi", "Horror"]

ave_merged_df = ave_merged_df.dropna()


matplotlib.style.use("seaborn")
ave_merged_df.plot(figsize = (15,8))
plt.xlabel("Year")
plt.ylabel("Average rating")
plt.title("Movie Genres : Average ratings per year")
plt.yticks([0,1,2,3,4,5])
plt.grid(True)
plt.show()
```
