# Random Forest Regression 

This is a very simple model that uses Random Forest Regression to fit the given data. All of the code is pretty straight forward.

Steps:
------

1. Import the Libraries
2. Read the dataset using pandas.read_csv as the data is a csv file
3. Extract the required features
4. Concatenate them into a feature vector
5. Call the Random Forest regressor 
6. Set the Number of estimators to 1200 , random_state to 9 and min_weight_fraction_leaf to 0.00001. These are some parameters that I found to be optimal considering the time taken to run and the best value of r2 score. 



### Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
```

### Loading the data

```python
dataset = pd.read_csv("../input/SolarEnergy/SolarPrediction.csv")
```

```python
dataset
```

### Extracting the necessary features

```python
temp = np.array(dataset["Temperature"]).reshape(-1,1)
press = np.array(dataset["Pressure"]).reshape(-1,1)
wd = np.array(dataset["WindDirection(Degrees)"]).reshape(-1,1)
humd = np.array(dataset["Humidity"]).reshape(-1,1)
speed = np.array(dataset["Speed"]).reshape(-1,1)
```

### Combining the features

```python
findata = np.concatenate((temp,press,humd,wd,speed),axis = 1)
```

```python
findata.shape
```

```python
dataset.shape
```

### Setting the target 

```python
y = np.array(dataset["Radiation"])
```

### Creating the Random Forest Regressor Instance

```python
regressor = RandomForestRegressor(n_estimators = 1200 , random_state = 9 , min_weight_fraction_leaf = 0.00001)
regressor.fit(findata,y)
```

### Plotting ground truth vs predicted values

```python
plt.plot(y)
plt.plot(regressor.predict(findata))
```

### Looking at the R2 score

```python
regressor.score(findata,y)
```
