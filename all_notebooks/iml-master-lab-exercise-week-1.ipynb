### Q2 - Print the mean, mode and median for the following list [5,7,2,4,5,6,4,5,6,4,3,5,6,5,3]

```python
# Write your code for Q2 and execute the cell
d = pd.Series([5,7,2,4,5,6,4,5,6,4,3,5,6,5,3])
print("Mean: ",d.mean())
print("Mode: ", d.mode())
print("Median: ", d.median())
```

### Q3 - Using **chickwits.csv**, print mean, median, 80% quantile value for column "weight". For the column "feed", print proportion and percentage values.

```python
# Write your code for Q3 and execute the cell
c1 = pd.read_csv("../input/chickwts.csv")
c1.head(10)
```

```python
w1 = c1["weight"]
display("Mean: ", w1.mean())
display("Median: ", w1.median())
display("80% value: ", w1.quantile(0.8))
display("Weight description:",w1.describe())
```

```python
f1 = c1["feed"]
print("Value Count:\n",f1.value_counts())
print('---------------------------')
print("Proportion:\n",f1.value_counts()/f1.size)
print('---------------------------')
print("Percentages:\n",f1.value_counts()/f1.size*100)
```

```python
sns.countplot(f1)
```

### Q5 - Using **orange.csv**, find the correlation between the columns "age" and "circumference". Also, generate a scatter plot for both the columns mentioned.

```python
# Write your code for Q5 and execute the cell
o = pd.read_csv("../input/orange.csv")
o1 = o[["age","circumference"]]
print(o1)
print("------------------------")
print("Correlation:\n",o1.corr())
```

```python
plt.scatter(o1["age"],o1["circumference"], alpha=0.5)
plt.title('Scatter plot')
plt.xlabel('age')
plt.ylabel('circumference')
plt.show()
```

### Q6 - Using **mtcars.csv**, generate a bar plot for the column "cyl".

```python
# Write your code for Q6 and execute the cell
m1=pd.read_csv("../input/mtcars.csv")
m1.head()
```

```python
cf = m1["cyl"].value_counts()
cd = cf.apply(pd.Series)
cd.columns = ["freq"]
print(cd)
```

```python
plt.bar(cd.index,cd["freq"])
plt.show()
```

### Q7 - Using **mtcars.csv**, generate a pie chart for the column "cyl".

```python
plt.pie(cd["freq"], labels=cd.index, startangle=90)
plt.show()
```
