### Questão 1 - Item C - Representação Gráfica
Para cada uma das variáveis, produza um ou mais gráficos, usando matplotlib, que descreva seu comportamento / caracteristica. Lembre-se que estes gráficos precisam ser compatíveis com a classificação da variável. 

```python
data = df['aeronave_motor_tipo'].value_counts().reset_index().head(5)

plt.figure(figsize=(7, 7))
plt.pie(x=data['aeronave_motor_tipo'], autopct='%1.1f%%', startangle=90 )
plt.title('Tipo de motor')
plt.legend(data['index'],  bbox_to_anchor=(1, 1))
plt.axis('equal')
plt.show()
```

```python
data = df['aeronave_motor_quantidade'].value_counts().reset_index().head(3)

plt.figure(figsize=(7, 7))
plt.pie(x=data['aeronave_motor_quantidade'], autopct='%1.1f%%', startangle=90 )
plt.title('Tipo de motor (top 3)')
plt.legend(data['index'],  bbox_to_anchor=(1, 1))
plt.axis('equal')
plt.show()
```

```python
data = df['aeronave_fabricante'].value_counts().reset_index().head(20)

plt.figure(figsize=(10, 15))
plt.style.use('ggplot')
plt.barh(y=data['index'][::-1], width=data['aeronave_fabricante'][::-1])
plt.title('Fabricantes mais frequentes')
plt.xlabel('Número de aeronaves')
plt.ylabel('Fabricante')
plt.show()
```

```python
data = df['aeronave_modelo'].value_counts().reset_index().head(50)

plt.figure(figsize=(10, 25))
plt.style.use('ggplot')
plt.barh(y=data['index'][::-1], width=data['aeronave_modelo'][::-1])
plt.title('Modelos mais frequentes')
plt.xlabel('Número de aeronaves')
plt.ylabel('Modelo')
plt.show()
```

```python
data = df['aeronave_tipo_veiculo'].value_counts().reset_index().head(3)

plt.figure(figsize=(7, 7))
plt.pie(x=data['aeronave_tipo_veiculo'], autopct='%1.1f%%', startangle=90 )
plt.title('Tipo de aeronave')
plt.legend(data['index'],  bbox_to_anchor=(1, 1))
plt.axis('equal')
plt.show()
```

```python
df['aeronave_operador_categoria'].value_counts().plot(kind='bar')


plt.xlabel("Categorias")
plt.ylabel("Quantidade por Categoria")
plt.title("Quantidade de Aeronaves por Categoria")
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

```python
df['aeronave_tipo_operacao'].value_counts().plot(kind='bar')

plt.xlabel("Operações")
plt.ylabel("Quantidade por operação")
plt.title("Quantidade de Aeronaves por operação")
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```
