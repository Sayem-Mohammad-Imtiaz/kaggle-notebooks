# Einleitung
Wir betrachten ein einfaches Beispiel für das MASCHINELLE LERNEN.

**Beispiel**: Es sollen die Preise für Häuser mit Hilfe des Computers bestimmt werden. Der Computer soll dabei die Entscheidung über den Preis treffen.

Ein einfacher Entscheidungsbaum soll dies verdeutlichen:

![First Decision Trees](http://i.imgur.com/7tsb5b1.png)

Der Entscheidungsbaum teilt die Häuser in zwei Kategorien, wobei der Preis durch vorherige durchschnittliche Verkaufspreise festgelegt wird.

Bekannte Daten wurden hier genutzt um die Häuser in zwei Gruppen zu teilen und um einen Verkaufspreis vorherzusagen.

Dieser Schritt wird als **fitting** bzw. **training** des Modells bezeichnet. Die Daten die zum **fitting** benutzt werden heißen **Trainingsdaten**. Im Anschluss kann nun das Modell für die **Vorhersage** mit neuen Daten genutzt werden.
