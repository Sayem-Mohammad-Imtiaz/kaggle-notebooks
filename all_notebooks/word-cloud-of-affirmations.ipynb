```python
from wordcloud import WordCloud, STOPWORDS
```

```python
affirmation_words = '' 
stop_words = set(STOPWORDS)
for i in data.Affirmation: 
    i = str(i) 
    separate = i.split() 
    for j in range(len(separate)): 
        separate[j] = separate[j].lower() 
      
    affirmation_words += " ".join(separate)+" "
```

```python
final_wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='black', 
                stopwords = stop_words, 
                min_font_size = 10).generate(affirmation_words)
```

```python
plt.figure(figsize = (10, 10), facecolor = None) 
plt.imshow(final_wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```

```python
for i in data["Tag"].unique():
    Love_words = ''
    stop_words = set(STOPWORDS)
    new_data = data[data['Tag'] == i]
    for i in new_data.Affirmation: 
        i = str(i) 
        separate = i.split() 
        for j in range(len(separate)): 
            separate[j] = separate[j].lower() 

        Love_words += " ".join(separate)+" "
    final_wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='black', 
                stopwords = stop_words, 
                min_font_size = 10).generate(Love_words)
    plt.figure(figsize = (10, 10), facecolor = None) 
    plt.imshow(final_wordcloud) 
    plt.axis("off") 
    plt.tight_layout(pad = 0) 

    plt.show()
    
```
