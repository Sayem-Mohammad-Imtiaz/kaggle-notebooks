# Model training

```python
from sklearn.preprocessing import LabelEncoder , MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
```

```python
sex_encoder = LabelEncoder()
sex_encoder.fit(df['Sex'])
df['Sex'] = sex_encoder.transform(df['Sex'])
```

```python
BP_encoder = LabelEncoder()
BP_encoder.fit(df['BP'])
df['BP'] = BP_encoder.transform(df['BP'])
```

```python
Cholesterol_encoder = LabelEncoder()
Cholesterol_encoder.fit(df['Cholesterol'])
df['Cholesterol'] = Cholesterol_encoder.transform(df['Cholesterol'])
```

```python
Drug_encoder = LabelEncoder()
Drug_encoder.fit(df['Drug'])
df['Drug'] = Drug_encoder.transform(df['Drug'])
```

```python
print(sex_encoder.classes_)
print('-'*10)
print(BP_encoder.classes_)
print('-'*10)
print(Cholesterol_encoder.classes_)
print('-'*10)
print(Drug_encoder.classes_)
```

```python
features = ['Age' ,	'Sex' , 'BP' ,	'Cholesterol' , 'Na_to_K']
features = df[list(features)]
label = df['Drug'].values
```

```python
x_train , x_test , y_train , y_test = train_test_split(features,label,test_size=.3)
```

```python
print("X train shape: ", x_train.shape)
print("Y train shape: ", y_train.shape)
print("X test shape: ", x_test.shape)
print("Y test shape: ", y_test.shape)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
print("Score is: ",model.score(x_test,y_test)*100)
```

```python
model = MultinomialNB(fit_prior=False)
model.fit(x_train,y_train)
print("Score is: ",model.score(x_test,y_test)*100)
```

```python
model = KNeighborsClassifier(n_neighbors=5)
model.fit(x_train,y_train)
print("Score is: ",model.score(x_test,y_test)*100)
```

```python
model = RandomForestClassifier()
model.fit(x_train,y_train)
print("Score is: ",model.score(x_test,y_test)*100)
```
