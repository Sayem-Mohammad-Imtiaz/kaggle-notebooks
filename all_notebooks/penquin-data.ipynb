```python
def bar_plot(variable):
    #****
    #    input: variable ex: 'Sex'
    #   output: bar plot & value count
    #****
    # get feature
    var = df[variable]
    #count number of categorical variable(value/sample)
    varValue=var.value_counts()
    
    #visualize
    plt.figure(figsize=(10,5))
    plt.bar(varValue.index,varValue)
    plt.xticks(varValue.index,varValue.index.values)
    plt.ylabel('Frequency')
    plt.title(variable)
    plt.show()
    print('{}: \n {}'.format(variable,varValue))
```

# Outlier Ditection

```python
df.columns
```

```python
def outlierr(df,variable):
    outlier_indices=[]
    
    for c in variable:
        #first quartile
        Q1=np.percentile(df[c],25)
        
        #third quartile
        Q3=np.percentile(df[c],75)
        
        #IQR
        IQR = Q3 - Q1
        
         
        #Detect outlier and their indeces
        outlier_list_col=df[(df[c]< Q1- IQR *1.5) | (df[c]> Q3 + IQR *1.5)].index
        
        #store indeces
        outlier_indices.extend(outlier_list_col)
    
    #Listenin içindeki aykırı değerlerden kaçar tane var onu hesaplatıyoruz.
    outlier_indices=Counter(outlier_indices)
    
    multiple_outliers= list(i for i, v in outlier_indices.items() if v>2)    
    
    print(multiple_outliers)
    
    return   multiple_outliers

        
    
```

Aykırı değer bulunmamaktadır.

```python
df.loc[outlierr(df,['Culmen Length (mm)', 'Culmen Depth (mm)', 'Flipper Length (mm)', 'Body Mass (g)', 'Delta 15 N (o/oo)','Delta 13 C (o/oo)'])]
```

```python
def outlier_boxplot(variable):
    g=sns.catplot(y=variable,x='Clutch Completion',data=df,kind="box", palette="Set1",linewidth=2.5,height=4, aspect=.7);
    
```

```python
variable=['Culmen Length (mm)', 'Culmen Depth (mm)', 'Flipper Length (mm)', 'Body Mass (g)', 'Delta 15 N (o/oo)','Delta 13 C (o/oo)']

for v in variable:
    outlier_boxplot(v)
```

Verilerden anlaşılacağı üzere sayısal değerlerin hepsinde eksik veri gözlemlenirken kategorik değişkende sadece Sex için eksik veri vardır. Eksik veri doldurma işlemi yapılırken sayısal ve kategorik değişkenlere ayrı bakılacaktır.

* Kategorik Değişkenin Doldurulmması(Sex değişkeni)

Sex değişkeni, Individual ID değişkeninde A2 ve A1 geçen gözlemlere göre incelenir. Gözlem sonucunda A2 geçenler Male olarak, A1 geçenler ise Female olarak sonraki adımda doldurulacaktır.

```python
male_sex=df["Individual ID"].str.contains("A2")
female_sex=df["Individual ID"].str.contains("A1")
```

Verisetinin toplam satır sayısı 344' dür. Bunu doğrulamak için erkek ve dişi adeti incelnip toplamına bakılır. Uygun şekilde doldurulur.

```python
print(Counter(df.Sex[male_sex]))
print(Counter(df.Sex[female_sex]))

df.Sex[male_sex]="MALE"
df.Sex[female_sex]="FEMALE"
```

```python
df.isnull().sum()
```

İncelendiğinde Sex değişkeni için eksik olan tüm değerler doldurulmuştur.

```python
msno.matrix(df,color=(0.5,0.3,0.7));
```

* Sayısal Değişkenlerin Doldurulması

Sayısal değişkenlerdeki eksik değerler doldurulurken değişkenler tür ve ada bilgisine göre gruplanarak doldurma işlemi yapılacaktır.

```python
def missingValue(variable):
    missing_list=[]
    
    for m in variable:
    
        a=df[df[variable].isnull()].index
        missing_list.extend(a)
        df[variable] = df[variable].fillna(df.groupby(['Species', 'Island'])[variable].transform('mean'))
   
    
    return missing_list
```

Eksin olan değerlerin index numarası görülür. Fonsiyonda gösterildiği gibi 'Species', 'Island', değişkenlerine göre gruplanır ve eksik olan değerler ortalama ile doldurulur.

```python
missingValue(['Culmen Length (mm)', 'Culmen Depth (mm)',
       'Flipper Length (mm)', 'Body Mass (g)', 'Delta 15 N (o/oo)',
       'Delta 13 C (o/oo)'])
```
