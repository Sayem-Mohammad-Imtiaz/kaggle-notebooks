The first question is, can an increase in sightings be accounted for?
Below is a histogram where the blue represents the number of reported events and the green represents the number of occurances.

```python
import datetime
import time
year_occurance = []
year_report =  []
for time in scrubbed['datetime']:
    if time[-5:] != '24:00':
        year_occurance.append(datetime.datetime.strptime(time, "%m/%d/%Y %H:%M").year)
for time in scrubbed['date posted']:
    year_report.append(datetime.datetime.strptime(time, "%m/%d/%Y").year)

plt.figure(1)
plt.hist(year_report, bins = 50, range = (1940,2015))
plt.hist(year_occurance, bins = 50, range = (1940,2015))
plt.ylabel('Amount')
plt.xlabel('Year')
#plt.figure(2)
#plt.ylabel('Number of reports')
#plt.xlabel('Year')
#plt.hist(year_report, bins = 30, range = (1940,2015))
```

As you can tell there has been an increase in both the number of both occurances and reportings in the last year. Now lets see if we can justify this. Below is a graph of the US population from 1940-2015. I chose to do only the US because a majority of reports came from the US. The Yscale is population divided by 1000.

```python
year_pop = []
year = []
for i in range(40000):
    if pop['Location'][i] == 'United States of America':
        year_pop.append(pop['PopTotal'][i])
        year.append(pop['Time'][i])
year_pop = year_pop[:65]
year = year[:65]
year_pop_fitted = []
for p in range(65):
    year_pop_fitted.append((year_pop[p])*0.0533333333)
plt.plot(year, year_pop)
plt.ylabel('1000 people')
plt.xlabel('Year')
```

It doesn't seem as though it will match up and that is made more apparent when the graph of population is put on top of that of the occurance/report graph.

```python
plt.figure(1)
plt.plot(year, year_pop_fitted,color = 'r')
plt.hist(year_report, bins = 50, range = (1940,2015))
plt.hist(year_occurance, bins = 50, range = (1940,2015))
```

As you can tell the increase in occurances does not match an increase in population, this was appairant from the start, but what if something else caused this increase, perhaps the internet. The National UFO Reporting Center has held a hotline since 1974, however its operations were expanded in 2005 so I will use this as the starting point of our graph. 

```python
plt.hist(year_report, bins = 10, range = (2005,2015))
plt.hist(year_occurance, bins = 10, range = (2005,2015))
plt.ylabel('Amount')
plt.xlabel('Year')
```

Alright, lets put another fitted population graph on top of this.

```python
year_pop_fitted_2 = []
for p in range(60):
    year_pop_fitted_2.append((year_pop[p])*0.0266666667)
year_pop_fitted_2 = year_pop_fitted_2[-10:]
plt.plot(year[-10:], year_pop_fitted_2,color = 'r')
plt.hist(year_report, bins = 11, range = (2005,2015))
plt.hist(year_occurance, bins = 11, range = (2005,2015))
plt.ylabel('Amount')
plt.xlabel('Year')
```

That line looks like a good fit for the graph accounting for the outliers past 2011 as shown in the graph below

```python
year_pop_fitted_3 = []
for p in range(55):
    year_pop_fitted_3.append((year_pop[p])*0.0266666667)
year_pop_fitted_3 = year_pop_fitted_3[-7:]
plt.plot(year[-10:][:7], year_pop_fitted_3,color = 'r')
plt.hist(year_report, bins = 7, range = (2005,2011))
plt.hist(year_occurance, bins = 7, range = (2005,2011))
plt.ylabel('Amount')
plt.xlabel('Year')
```

So can we account for an increase? Yes, surprisingly, taking into account outliers and the emergance of the internet the seemingly sudden increase in reports makes sense.
