```python
# import libraries

import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
# assigning files to their respective variables

athlete_events = pd.read_csv('../input/120-years-of-olympic-history/athlete_events.csv')
country_def = pd.read_csv('../input/120-years-of-olympic-history/country_definitions.csv')
```

```python
# let's take a look on data

athlete_events.head(5)
```

```python
athlete_events.info()
```

```python
# corelation beetwen medals per country

medals_per_country = athlete_events[['NOC', 'Medal']]
medals_per_country = medals_per_country.loc[medals_per_country['Medal'] != 'NaN'].groupby('NOC').count().sort_values(by='Medal',ascending=False)
medals_per_country = pd.merge(medals_per_country,
                              country_def,
                              how="inner",
                              on='NOC')


medals_per_country = medals_per_country[['region', 'Medal']].groupby('region').sum().sort_values(by='Medal', ascending=False)
medals_per_country = medals_per_country.reset_index()

px.bar(data_frame=medals_per_country[0:20], x='region', y='Medal')
```

```python
# corelation between man and women medals

medals_per_sex = athlete_events[['Medal', 'Sex']]
medals_per_sex = medals_per_sex.loc[medals_per_sex['Medal'] != 'NaN'].groupby('Sex').count()
medals_per_sex = medals_per_sex.reset_index()
sns.catplot(data=medals_per_sex, x='Sex', y='Medal', kind='bar')
```

We see that men wins more medals than women

```python
# the trend for the number of participants

participants_by_event = athlete_events[['ID', 'Year']]
participants_by_event = participants_by_event.loc[participants_by_event['ID']].groupby('Year').count()
participants_by_event = participants_by_event.reset_index()
px.bar(data_frame=participants_by_event, x='Year', y='ID')
```

The trend for the number of participants is growing.

```python
# corelation between age and medals won

medals_per_age = athlete_events[['Age', 'Medal']]
medals_per_age = medals_per_age.loc[medals_per_age['Medal'] != 'NaN'].groupby('Age').count()
medals_per_age = medals_per_age.reset_index()
px.bar(data_frame=medals_per_age, x='Age', y='Medal')
```

There is slightly positively skewed distribution. The most medals wins participants in age 23 and 24.

```python
# corelation between height and medals won

medals_per_height = athlete_events[['Height', 'Medal']]
medals_per_height = medals_per_height.loc[medals_per_height['Medal'] != 'NaN'].groupby('Height').count()
medals_per_height = medals_per_height.reset_index()


# let's assume that tall people are 175cm and highier
tall_people = medals_per_height.loc[medals_per_height['Height'] >= 175.0].sum()
tall_people = tall_people['Medal']

short_people = medals_per_height.loc[medals_per_height['Height'] < 175.0].sum()
short_people = short_people['Medal']

print('Tall people medals: ', int(tall_people))
print('Short people medals: ', int(short_people))

# let's look at corelation
px.bar(data_frame=medals_per_height, x='Height', y='Medal')
```

As we calculated the most medals wins tall people.

```python
# which sports are dominated by short people

short_people_sports = athlete_events[['Height', 'Sport']]

# previously we assume that tall people are 175 cm and highier so short people will be below 175cm 
short_people_sports = short_people_sports.loc[short_people_sports['Height'] < 175.0].groupby('Sport').count().sort_values(by='Height', ascending=False)
short_people_sports = short_people_sports.reset_index()

# let's look at corelation
px.bar(data_frame=short_people_sports[0:10], x='Sport', y='Height')
```

The most dominated sports by short people are Gymnastic and Athletics.
