```python
df= pd.read_csv("/kaggle/input/the-human-freedom-index/hfi_cc_2018.csv")
df
```

```python
cualitativas = list()
cualitativas.append(df.columns[1:4])
cualitativas
```

```python
cuantitativas = list()
cuantitativas.append(df.columns[0])
cuantitativas.extend(df.columns[4:])
cuantitativas
```

```python
rangos = list()
for i in cuantitativas:
            rango= np.max(df[i])- np.min(df[i])
            rangos.append(rango)
rangos
```

```python
medias=[]
for i in cuantitativas:
    media=df[i].mean()
    medias.append(media)
medias
```

```python
import scipy
from scipy import stats 
modas=[]
for i in cualitativas:
    modas=stats.mode(df[i])
modas[0][0]
```

```python
varianzas=list()
for i in cuantitativas:
    varianza=np.var(df[i])
    varianzas.append(varianza)
varianzas
```

```python
desviaciones_estandar=[]
for i in cuantitativas:
    desviacion_estandar=np.std(df[i])
    desviaciones_estandar.append(desviacion_estandar)
desviaciones_estandar
```
