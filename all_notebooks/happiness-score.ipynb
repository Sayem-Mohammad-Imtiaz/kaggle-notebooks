```python
data_2020 = pd.read_csv('/kaggle/input/world-happiness-report-preprocessed/2020_report.csv')
```

```python
data_2020 = data_2020.drop(['country'],axis = 1)
```

```python
data_2020
```

```python
X = data_2020.iloc[:, 1:].values
y = data_2020.iloc[:, 0].values
```

Data Preprocessing

```python
print(X)
```

```python
y = y.reshape(len(y),1)
```

```python
print(y)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [7])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

Regression Models

```python
from sklearn.linear_model import LinearRegression
reg1 = LinearRegression()
reg1.fit(X_train, y_train)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X_train)
X_testpoly = poly_reg.transform(X_test)
reg2 = LinearRegression()
reg2.fit(X_poly, y_train)
```

```python
from sklearn.svm import SVR
reg3 = SVR(kernel = 'linear')
reg3.fit(X_train, y_train)
```

```python
from sklearn.tree import DecisionTreeRegressor
reg4 = DecisionTreeRegressor(random_state = 0)
reg4.fit(X_train, y_train)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg5 = RandomForestRegressor(n_estimators = 1000, random_state = 0)
reg5.fit(X_train, y_train)
```

```python
ypred1 = reg1.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(ypred1,y_test)
```

```python
ypred2 = reg2.predict(X_testpoly)
r2_score(ypred2,y_test)
```

```python
ypred3 = reg3.predict(X_test)
r2_score(ypred3,y_test)
```

```python
ypred4 = reg4.predict(X_test)
r2_score(ypred4,y_test)
```

```python
ypred5 = reg5.predict(X_test)
r2_score(ypred5,y_test)
```
