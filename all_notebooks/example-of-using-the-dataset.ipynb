```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

```

```python
path="/kaggle/input/using-geodata-in-python/ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp"
import geopandas as gpd
#to draw the contours of the countries
world = gpd.read_file(path)
path2="/kaggle/input/serial-killers-wikipediadata/World_serial_killer.csv"
my_data=pd.read_csv(path2)
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns
df_count=pd.DataFrame({'Count': my_data["Country"].value_counts().sort_values(ascending = False)})
label=list(df_count.index)
fig=plt.figure(figsize=[20,8])
fig.patch.set_facecolor('#E0E0E0')
fig.patch.set_alpha(0.7)
plt.title("Number of serial killers per country (from Wikipedia)",size=16)
plt.bar(range(0,len(label)),df_count["Count"]
           ,edgecolor='black',color="#8891a1")
plt.xticks(range(0,len(label)),label,rotation=90,size=13)
plt.ylabel("Nombre")
plt.grid()
plt.show()
```

```python
df_count=pd.DataFrame({'Count': my_data["name"].value_counts().sort_values(ascending = False)})
world['count'] = world.name.map(df_count['Count'])
ax = world.plot(column = world["count"], cmap = 'Reds' ,
                legend = True, scheme="percentiles", edgecolor = 'black', linewidth = 0.3, figsize = (15,18),
                missing_kwds={ 'color': "#eceeeb",'edgecolor': "#6b6b6b", 'hatch': "///",'label': "Missing values"})
plt.title('Number of serial killers in each country.', fontsize = 16)
plt.grid()
```

```python
europe=world.loc[world["continent"]=="Europe"]
ax = europe.plot(column = europe["count"], cmap = 'Reds', legend = True,
                 edgecolor = 'black', linewidth = 0.3, figsize = (12,8)
                 ,missing_kwds={ 'color': "#eceeeb",
                                'edgecolor': "#6b6b6b", 
                                'hatch': "///",'label': "Missing values"})
plt.title('Number of serial killers in each country of Europe Zone.', fontsize = 16)
plt.xlim(-15,30)
plt.ylim(32,70)
plt.legend()
plt.grid()
plt.tight_layout()
```

```python
pop=world[["name","pop_est"]].copy()
test=df_count.copy()
test["pop_millions"]=np.nan

for i in test.index:
    p=pop.loc[pop["name"]==i]["pop_est"].values[0]
    test.loc[i,"pop_millions"]=round(p/1000000,2)
    
test["prevalence per millions"]=test["Count"]/test["pop_millions"]

test.head()
```

```python
world['prevalence'] = world.name.map(test['prevalence per millions'])

ax = world.plot(column = world['prevalence'], cmap = 'Reds' ,legend=True
                , scheme="percentiles", edgecolor = 'black', linewidth = 0.3, figsize = (18,15),
                missing_kwds={ 'color': "#eceeeb",'edgecolor': "#6b6b6b", 'hatch': "///",'label': "Missing values"})
plt.title('Prevalence of serial killers, per million inhabitants.', fontsize = 16)
plt.grid()

plt.tight_layout()
```

```python
europe=world.loc[world["continent"]=="Europe"]
europe=europe.loc[europe["name"]!="Iceland"]
ax = europe.plot(column = europe['prevalence'], cmap = 'Reds', legend = True,
                 edgecolor = 'black', linewidth = 0.3, figsize = (12,8)
                 ,missing_kwds={ 'color': "#eceeeb",
                                'edgecolor': "#6b6b6b", 
                                'hatch': "///",'label': "Missing values"})
plt.title('Prevalence of serial killers in each country of Europe Zone.\n per Million', fontsize = 16)
plt.xlim(-15,30)
plt.ylim(32,70)
plt.legend()
plt.grid()
plt.tight_layout()
```
