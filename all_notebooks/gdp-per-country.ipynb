Examining all the countries using pycountry

```python
list(pycountry.countries)
```

```python
countries_list = [ x.alpha_3 for x in pycountry.countries]
result = data['Country Code'].isin(countries_list)
n=[]
for x in range(len(result)):
    if(result[x]==False):
        n.append(x)
len(n)
```

```python
only_countries_df = data.drop(n,axis=0)
only_countries_df.count()
```

**** Analyzing by the year

```python
def get_the_largest_GDP(year):
    d = only_countries_df.nlargest(10,year)[year].tolist()
    c = only_countries_df.nlargest(10,year)['Country Name'].tolist()
    plt.figure(figsize=(7,7))
    plt.title(label= "GDP Contribution of the top 10 countries in " + year, fontsize=20, fontweight=20)
    plt.pie(d,labels=c, startangle=80)
    plt.show()
```

```python
get_the_largest_GDP('1960')
get_the_largest_GDP('1970')
get_the_largest_GDP('1980')
get_the_largest_GDP('1990')
get_the_largest_GDP('2000')
```

# Country Analysis

```python
only_countries_df.set_index('Country Name', inplace= True)
only_countries_df.loc['Aruba'][3:64].describe
```

```python
print(only_countries_df.loc['France'][3:64].describe)
```

```python
plt.figure(figsize=(30,22))
plt.title(label= 'France vs. Canada GDP Analysis', fontsize=30, fontweight=20)
plt.ylabel('GDP in terms of $')
plt.xlabel('Years',fontsize=25)
plt.plot(np.arange(1960,2021),only_countries_df.loc['France'][3:64],label='France')
plt.plot(np.arange(1960,2021),only_countries_df.loc['Canada'][3:64],label='Canada')
plt.xticks(np.arange(1960,2021),np.arange(1960,2021), rotation=45)
plt.grid()
```

```python
def get_the_country_plot(country):
    print(country + " GDP Summary")
    print()
    print('Mean : ', end= '')
    print(only_countries_df.loc[country][3:64].mean())
    print('Total GDP Until Now: ', end='')
    print(only_countries_df.loc[country][3:64].sum())
    print('Maximum GDP: ', end= '')
    print(only_countries_df.loc[country][3:64].max())
    print('Minimum GDP: ', end= '')
    print(only_countries_df.loc[country][3:64].min())
    plt.figure(figsize=(30,22))
    plt.title(label= country + ' GDP Analysis', fontsize=30, fontweight=20)
    plt.ylabel('GDP in terms of $')
    plt.xlabel('Years',fontsize=25)
    plt.plot(np.arange(1960,2021),only_countries_df.loc[country][3:64])
    plt.xticks(np.arange(1960,2021),np.arange(1960,2021), rotation=45)
    plt.grid()
```

```python
get_the_country_plot('Ireland')
```

```python
get_the_country_plot('Thailand')
```

```python
get_the_country_plot('Japan')
```
