```python
# Importing Pandas As pd
import pandas as pd
# importing numpy as np
import numpy as np

# Reading Data Frame
df =  pd.read_csv("../input/bike-buyers/bike_buyers.csv")
df.head(6)
```

```python
# Replacing Marital Status null values With "No Data"
df['Marital Status'].fillna("No Data", inplace=True)

# Replacing Gender null values  With "No Data"
df['Gender'].fillna("No Data", inplace=True)

# Replacing Home Owner null values  With "No Data"
df['Home Owner'].fillna("No Data", inplace=True)
```

```python
# Importing Matplotlib
import matplotlib.pyplot as plt
# Import Seaborn as sns
import seaborn as sns

# Sizing The Plot Size
sns.set(rc={'figure.figsize':(20,15)})

# Using Matplotlib for Subploting
fig, axes = plt.subplots(3,1)

# Subploting Data Showing Income According to Region and Professin 
sns.kdeplot(data = df[df['Region'] == 'Europe'], x = 'Income', hue = 'Occupation', ax = axes[0])
sns.kdeplot(data = df[df['Region']=='North America'],x='Income', hue='Occupation', ax=axes[1])
sns.kdeplot(data = df[df['Region']=='Pacific'],x='Income', hue='Occupation', ax=axes[2])
```

```python
# Finding Income Null in Europe
df[pd.isnull(df['Income']) & (df['Region'] == 'Europe')]
```

```python
# Finding The Median Value Of Manual Occupation in Europe(ei=Europe income).
ei = df[(df['Region'] == 'Europe') & (df['Occupation'] == 'Manual')]['Income'].median()

# Making The Income Null in Europe a list(einl=europe income null value list).
einl = df[pd.isnull(df['Income']) & (df['Region'] == 'Europe')].index

# Repalcing The Europe Income Null Value with median
for item in einl:
    df.iloc[item, df.columns.get_loc('Income')] = ei
```

```python
# Finding Income Null in North America
df[(pd.isnull(df['Income'])) & (df['Region']=='North America')]
```

```python
# Finding The Mode Value Of Income Accourding to Bachelor Occupation in North America
# (nai=north america income).
nai = df[(df['Region'] == 'North America') & (df['Occupation'] == 'Professional')]['Income'].mode()

#  Repalcing The North America Income Null Value with mode
df.iloc[509, df.columns.get_loc('Income')] = nai
```

```python
# Finding Income Null in Pacific
df[(pd.isnull(df['Income'])) & (df['Region']=='Pacific')]

# Finding The Mode Value Of Income Accourding to Clerical & Skilled Manual Occupation in Pacific
# (pi=Pacific income).
pi = df[(df['Region'] == 'Pacific') & (df['Occupation'] == 'Clerical')]['Income'].mode()
pi2 = df[(df['Region'] == 'Pacific') & (df['Occupation'] == 'Skilled Manual')]['Income'].mode()

#  Repalcing The Pacific Income Null Value with mode
df.iloc[301, df.columns.get_loc('Income')] = pi
df.iloc[441, df.columns.get_loc('Income')] = pi2
```

```python
# Finding Children Null in Data
df[pd.isnull(df['Children'])]
```

```python
# Finding Effect On children Data by Region and Home Owner Ship.
sns.lineplot(data=df, y='Children', x='Region', hue = 'Home Owner')
```

```python
# Finding The Mode Value Of Having Children Accourding to Region and Homer Owner in North America
# (cna= Children north america).
cna = df[(df['Region'] == 'North America') & (df['Home Owner'] == 'Yes')]['Children'].mode()
cna2 = df[(df['Region'] == 'North America') & (df['Home Owner'] == 'No')]['Children'].mode()

# Making The Children Null in North America a list(cnnal=Children North America value list).
cnnal = df[pd.isnull(df['Children']) & (df['Region'] == 'North America') & (df['Home Owner'] == 'Yes')].index

#  Repalcing The Children Null Value in North America with mode with home.
for item in cnnal:
    df.iloc[item, df.columns.get_loc('Children')] = cna
# Repalcing The Children Null Value in North America with mode with out home.
df.iloc[688, df.columns.get_loc('Children')] = cna2
df.iloc[960, df.columns.get_loc('Children')] = cna2

# Finding The Mode Value Of Having Children Accourding to Region and Homer Owner in europe.
# (ce= Children Europe).
ce = df[df['Region'] == 'Europe'] ['Children'].mode()

#  Repalcing The Having Children Null Value with mode.
df.iloc[117, df.columns.get_loc('Children')] = ce
df.iloc[217, df.columns.get_loc('Children')] = ce

# Finding The Mode Value Of Having Children Accourding to in Pacific
# (cp= Children Pacific).
cp = df[(df['Region'] == 'Pacific') & (df['Age'] == 34)]['Children'].mode()

# Replaceing data 
df.iloc[386, df.columns.get_loc('Children')] = cp
```

```python
# Finding Effect On Age Data by Region and Home Owner Ship.
sns.lineplot(data=df, y='Cars', x='Region', hue = 'Home Owner')
```

```python
# Finding The Mode Value Of Car Accourding to Region and Homer Owner in Europe
# (care= Car In Europe).
care = df[(df['Region'] == 'Europe') & (df['Home Owner'] == 'Yes')]['Cars'].mode()

# Making The Car Null in Europe a list(carenl=Cars europe null list).
carenl = df[pd.isnull(df['Cars']) & (df['Region'] == 'Europe') & (df['Home Owner'] == 'Yes')].index

#  Repalcing The Car Null in Europe with mode with home.
for item in carenl:
    df.iloc[item, df.columns.get_loc('Cars')] = care

## Finding The Mode Value Of Car Accourding to Region and Homer Owner in North America.
# (carna= Car In North America).
carna = df[(df['Region'] == 'North America') & (df['Home Owner'] == 'Yes')]['Cars'].mode()

# Repalcing The Car Null in North America with mode with home.
df.iloc[511, df.columns.get_loc('Cars')] = carna
df.iloc[933, df.columns.get_loc('Cars')] = carna

## Finding The Mode Value Of Car Accourding to Region and Homer Owner in North America.
# (carna2 = Car In North America).
carna2 = df[(df['Region'] == 'North America') & (df['Home Owner'] == 'No')]['Cars'].mode()

# Repalcing The Car Null in North America with mode with home.
df.iloc[561, df.columns.get_loc('Cars')] = carna2
df.iloc[615, df.columns.get_loc('Cars')] = carna2
```

```python
#  Finding Age Null in Data
df[pd.isnull(df['Age'])]
```

```python
# Finding Effect On Age Data by Region and Home Owner Ship and Purchased Bike.
sns.lineplot(data=df, y='Age', x='Region', hue = 'Home Owner')
```

```python
# Finding The Mode Value Of Age Accourding to Region,Purchased Bike and Homer Owner in Europe
# (ae= age In Europe).
ae = df[(df['Region'] == 'Europe') & (df['Home Owner'] == 'Yes') & (df['Purchased Bike'] == 'Yes')]['Age'].mode()
ae2 = df[(df['Region'] == 'Europe') & (df['Home Owner'] == 'Yes') & (df['Purchased Bike'] == 'No')]['Age'].mode()

# Repalcing The Age Null in Europe with mode.
df.iloc[9, df.columns.get_loc('Age')] = ae
df.iloc[98, df.columns.get_loc('Age')] = ae
df.iloc[225, df.columns.get_loc('Age')] = ae2

## Finding The Mode Value Of Age Accourding to Region and Homer Owner in Pacific.
# (ap= Age In Pacific).
ap = df[(df['Region'] == 'Pacific') & (df['Home Owner'] == 'Yes') & (df['Purchased Bike'] == 'No')]['Age'].mode()

# Repalcing The Age Null in Pacificwith mode.
df.iloc[371, df.columns.get_loc('Age')] = ap

# Finding The Mode Value Of Age Accourding to Region,Purchased Bike and Homer Owner in Europe
# (ana= age In North America).
ana = df[(df['Region'] == 'North America') & (df['Home Owner'] == 'Yes') & (df['Purchased Bike'] == 'Yes')]['Age'].mode()
ana2 = df[(df['Region'] == 'North America') & (df['Home Owner'] == 'No') & (df['Purchased Bike'] == 'No')]['Age'].mode()

# Repalcing The Age Null in North America with mode.
df.iloc[554, df.columns.get_loc('Age')] = ana
df.iloc[986, df.columns.get_loc('Age')] = ana
df.iloc[688, df.columns.get_loc('Age')] = ana2
df.iloc[770, df.columns.get_loc('Age')] = ana2
```
