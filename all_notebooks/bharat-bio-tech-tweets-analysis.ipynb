# Text cleaning

```python
df = df.drop(['username','name','language'],axis = 1)
```

```python
df.head(10)
```

```python
import string
def clean_text(text):
    '''Make text lowercase, remove text in square brackets,remove links,remove punctuation
    and remove words containing numbers.'''
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    return text
```

```python
df.tweet = df.tweet.apply(lambda x:clean_text(x))
```

# Creating targets( POSITIVE, NEGATIVE, NEUTRAL) using VADER

```python
sentiments = []
for count,tweet in enumerate(df.tweet):
#     print(tweet)
    sentiments.append(vader.polarity_scores(tweet))
    print(count,vader.polarity_scores(tweet))
```

```python
len(sentiments)
```

```python
target = []
for count,sentiment in enumerate(sentiments):
    print(count)
    if sentiment['compound'] >= 0.05: 
        print("Positive")
        target.append('POSITIVE')
    elif sentiment['compound'] <= -0.05 : 
        print("Negative")
        target.append('NEGATIVE')
    else: 
        print("Neutral")
        target.append('NEUTRAL')
```

```python
print(target)
```

```python
target = pd.DataFrame(target, columns=['target'])
target
```

```python
type(target)
```

```python
df = pd.concat([df, target], axis = 1)
```

```python
w2v_model = gensim.models.word2vec.Word2Vec(size=300, 
                                            window=7, 
                                            min_count=10, 
                                            workers=8)
```

```python
%%time
documents = [_text.split() for _text in df.tweet_without_stopwords] 
```

```python
w2v_model.build_vocab(documents)
```

```python
words = w2v_model.wv.vocab.keys()
vocab_size = len(words)
print("Vocab size", vocab_size)
```

```python
w2v_model.train(documents, total_examples=len(documents), epochs=8)
```
