### DEFAULT and FLEXIBLE ARGUMENTS

```python
def haha(*args):
    for i in args:
        print(i)
haha(2023,334,3555,"sdbsdjf")
```

```python
def hehe(**kwargs):
    for key,value in kwargs.items():
        print(key,":",value)
hehe(yaban = "minnak", canan = "cinnak", hamur = "abi")
```

### ITERATORS

```python
liste1 = ["yankı","mancınık",43,65]
liste2 = [43,"uyu",56,"yan"]
z = zip(liste1,liste2)
zlistesi = list(z)
print(zlistesi)
```

### LIST COMPREHENSİON

```python
hehey = [3,4,6,7,89,9]
hudey = [i+3 for i in hehey]
print(hudey)
```

```python
halay = [3,-10,301,34,6,7,8,23]
yalan = [i-10 if i>10 else i+20 if i<1 else i for i in halay]
print(yalan)
```
