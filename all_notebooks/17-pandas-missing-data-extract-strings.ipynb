```python
# 만약 Age 컬럼의 평균값으로 빈값을 채운다면?
temp= train.copy()  # 테스트를 위해 복사본 작성
temp.Age = temp.Age.fillna(temp.Age.mean())
```

나이와 생존여부가 직접적인 관계를 보이지 않지만 빈값 177개를 평균 나이로 채우기에는 나이 데이터 값이 다양함.. 승객의 연령대를 추측할 방법이 없을까? 이름에서 힌트를 얻어보자!

```python
# Name 컬럼에서 이니셜만 추출
train['Initial'] = train.Name.str.extract('([A-Za-z]+)\.')  # period(.) 바로 앞의 문자열 추출
```

```python
train.Initial.unique()
```

```python
# Age 컬럼의 빈값을 이니셜별 평균 나이로 입력하기
train['Age'] = train.groupby('Initial').Age.apply(lambda x: x.fillna(x.mean()))
```

```python
# Cabin에서 첫글자(이니셜)만 추출하기
train.Cabin.str[0]
```

```python
# Name 컬럼을 분할
train.Name.str.split(', ')  # 콤마 및 한 칸 띄어쓰기 단위로 분할
```

```python
# 성(Last name) 출력하기 - 분할해서 생긴 리스트의 첫 번째 요소
train.Name.str.split(', ').str[0]
```

```python
# 성(Last name) 출력하기 -- (1) 첫번째 문자열로 출력
train.Name.str.extract('([A-Za-z]+)')
```

```python
# 성(Last name) 출력하기 -- (2) 콤마(,) 앞 문자열로 출력
train.Name.str.extract('([A-Za-z]+)\,')
```

```python
# Cabin에서 첫글자(이니셜)만 추출하기
train.Cabin.str.extract('([A-Za-z])')
```
