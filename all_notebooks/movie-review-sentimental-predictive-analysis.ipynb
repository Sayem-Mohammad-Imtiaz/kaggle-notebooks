# **What is IDBM Movie Review is ?**



* IDBM movie review (comments) is used to review movies by each watcher to provide some review about movie. so that next watcher will check the review ratings.
* Based on those review rating watcher can decide whether he want to watch the moview or not. Our timing is very valuable this review rating will be helpful.
* We can't go and check each and every review whether its +ve or -ve.But we can develop a model which it will classify +ve and -ve reviews as a moview rating.

* Missing value ration count and visualize the missing values count.

```python
# This logic will help to find missing ratio.
missing_ratio=(dataset.isna().sum()/len(dataset))*100.0

# To know all missing value ratio for each attribute....
print(missing_ratio.sort_values(ascending=False))
```

```python
# Visualising the missing ration in barplot

import seaborn as sb
import matplotlib.pyplot as plt

sb.barplot(x=missing_ratio.index,y=missing_ratio)
plt.title("Missing Value ratio in %")
plt.xlabel('Missing value attributes')
plt.ylabel('Missing value in percentage')
plt.show()
```

* We don't have any null values as per above outcome. So we can go ahead safely.
* As per missing value ratio and visualisation its clear that we dont have any missing values. If we have missing values then picture for ratio and barplot would be different. Please have a look on below picture.
