# Task Details

Here’s a list of questions we are going to answer to better understand what has been published regarding COVID-19 medical care:
What has been published about medical care? What has been published concerning surge capacity and nursing homes? What has been published concerning efforts to inform the allocation of scarce resources? What do we know about personal protective equipment? What has been published concerning alternative methods to advise on disease management? What has been published concerning the processes of care? What do we know about the clinical characterization and management of the virus?

With a better understanding of these questions, it will not only help researchers and scientists in academia, but also provide frontline healthcare workers and decision-makers with a clear understanding of the medical environment during the pandemic.

In the first phase, we explored data integrity, accessibility, research institutions, and publication timeline, and provided an overview of the literature. Next, we constructed a Keywords-Article Matrix, and used BERT SQuAD based Question-Answer Model as tools to help the community better and quickly understand the topics.

This report includes:

1. Data Integrity and Accessibility
2. Research Institutions and Publication Timeline
3. Literature Overview
4. Keywords-Article Matrix
6. Question-Answer Model

# 4. Keywords-Article Matrix

We constructed a Keywords-Article Matrix to understand the articles’ properties for the 16 key findings based on the task requirement. We used the frequency of words that were repeated in the combined titles and abstracts to extract meaningful keywords.

<img src="https://storage.googleapis.com/kagglesdsdata/datasets/604783/1084859/4.1.png?GoogleAccessId=web-data@kaggle-161607.iam.gserviceaccount.com&Expires=1587687001&Signature=gRdDFNZhUhQFhJe49XXs12920jHq4Ln7cR01W%2BRYVoWEwBNZlomahvAUQI%2BH2f6o3WywlWKunWIMTPQ97xsEG8r5FYlamcU5foBfBTUleEsv2Hp%2B8znjiOlRqzP6vzfFBig4R1gk5ubhHrmBJI6OyIuWMVDJHGu0a0KIwwLzy4Mrlj68lPTlB%2FlOxj1XBa7XZclhwIa9dJEeop3WAuvo6%2FsBsiAQ3qygxGDTUipQ0aShODrfy%2FiRvG8IWaBIohyCWyTT7g%2BMfOcgaIiied3p8UAkijyvSyLPGxXwowap7AWpShvaAnnzSWHf0pBp4%2BX6Ij5KOppsPZU6L69FkXtL2Q%3D%3D" width="450" />

We, then, come up with 16 sets of keywords based on the research review, brainstorm, and medical care professional consultation. For example, the keywords for the 1st key understanding, resources to support skilled nursing facilities and long term care facilities, are skilled nursing, long term care (or LTC), facility, health care, etc.
What’s more, the study of  “Perceptions of hand hygiene practices in China” used “a qualitative survey design based on in-depth interviews with 25 key hospital and public health staff in eight hospitals selected by the Chinese Ministry of Health.” Our Keywords-Article Matrix revealed that this article was about “the best practices and critical challenges and innovative solutions and technologies in hospital flow and organization, workforce protection, workforce allocation, community-based support resources, payment, and supply chain management to enhance capacity, efficiency, and outcomes”. It can also be about “Efforts to define the natural history of the disease to inform clinical care, public health interventions, infection prevention control, transmission, and clinical trials.”

# 5. Question-Answer Model

We use BERT and Stanford Question Answering Dataset (SQuAD) to build a Q&A model to help quickly find the possible answer from the articles for any related questions.
For instance, When you ask the model a question about “What is related to medical health care of covid-19 or coronavirus?” The Question-Answer model shows you a list of the related sentence from the articles with its ID. One of the returns for the question is “ID: 1019 Sentence: The impact of a 45-day delay in the epidemic due to early and accurate airport screening, allowing an effective vaccine to be developed and distributed, is also depicted.”
