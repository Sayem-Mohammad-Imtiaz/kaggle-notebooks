**Obtén una lista con los atributos cuantitativos y otra lista con los atributos cualitativos.**

```python
df = df.replace("-",np.nan)
df = df.replace(" ",np.nan)
```

```python

import scipy
from scipy import stats
for i in df.columns:
    df[i]  = df[i].replace(np.nan,stats.mode(df[i])[0][0])
```

```python
lista =[]
for i in df.iloc[0].values:
    try:
        if type(float(i))==float:
            lista.append("cuantitativo")
    except:
        lista.append("cualitativo")
lista
```

```python
for i,j in zip(lista,df.columns):
    if i == "cuantitativo":
        df[j] = df[j].astype("float")
    else:
        df[j]= df[j].astype("category")
```

**1.Agrega el calculo del coeficiente de asimetría y curtosis a cada columna cuantitativa**


```python
from scipy.stats import kurtosis
from scipy.stats import skew

for i in range(4,120):
    curtosis = kurtosis(df[df.columns[i]].values)
    
    print(df[df.columns[i]].name,"kurtosis=",curtosis)

print("-------------------"*10)

for i in range(4,120):
    asimetria =skew(df[df.columns[i]].values)
    print(df[df.columns[i]].name,"skew=",asimetria)
```

**2.Calcular el porcentaje de datos dentro de 1 desviación standard y 2 desviaciones standard**


```python
for i in range(4,120):
    media = np.mean(df[df.columns[i]].values) 
    desviacion = np.sqrt(np.var(df[df.columns[i]].values))
    rango1=media-desviacion
    rango2=media+desviacion
    rango_total=(len(df[df[df.columns[i]].between(rango1,rango2)]))/len(df[df.columns[i]].values)*100
    print("El porcentaje de datos dentro de 1 desviación estándar", df[df.columns[i]].name, "es del", rango_total, "%")
```

```python
for i in range(4,120):
    media = np.mean(df[df.columns[i]].values) 
    desviacion2 = (np.sqrt(np.var(df[df.columns[i]].values)))*2
    rango1=media-desviacion2
    rango2=media+desviacion2
    rango_total=(len(df[df[df.columns[i]].between(rango1,rango2)]))/len(df[df.columns[i]].values)*100
    print("El porcentaje de datos dentro de 2 desviaciones estándar", df[df.columns[i]].name, "es del", rango_total, "%")
```

**3.Gráfica las distribuciones de cada atributo e indica si el coeficiente de asimetría y curtosis son negativos o positivos.**


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in ["pf_rol_civil","pf_rol_criminal","pf_ss_disappearances_organized"]:
    #Gráfica de la columna
    sns.distplot(df[i],hist=False, color="black")
    
    #Cálculo media y desviacion
    media = np.mean(df[i])
    desviacion = np.sqrt(np.var(df[i])
    
    #Graficar las rectas con los valores de la media y la desviación
    #Uso la función axvline porque solo quiero rectas verticales
    
    plt.axvline(x=media, ymin=-2, ymax=2,color= "blue") 
    plt.axvline(x=media-desviacion, ymin=-2, ymax=2,color= "orange")
    plt.axvline(x=media+desviacion, ymin=-2, ymax=2,color= "orange")
    plt.axvline(x=media-(desviacion*2), ymin=-2, ymax=2,color= "orange")
    plt.axvline(x=media+(desviacion*2), ymin=-2, ymax=2,color= "orange")
    
    #Coeficiente de asimetria
    
    if skew(df[i]) > 0:
        asimetria = "Coeficiente de asimetría positivo"
    elif skew(df[i]) == 0:
        asimetria = "Coeficiente de asimetría de distribución normal"
    else:
        asimetria = "Coeficiente de asimetría negativo"
        
   #Curtosis  

    if kurtosis(df[i]) > 0:   
        curtosis = "Leptocúrtica. Decaimiento rápido, cola ligera"
    elif kurtosis(df[i]) == 0: 
        curtosis = "Mesocúrtica. Curva normal"
    else:
        curtosis = "Paticúrtica. Decaimiento lento."
        
    plt.title(asimetria + " - " + curtosis)
    
    plt.show()
   
```
