```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
degree=pd.read_csv('/kaggle/input/bachelor-degree-majors-by-age-sex-and-state/Bachelor_Degree_Majors.csv')
```

```python
degree.head()
```

```python
degree.tail()
```

```python
degree.info()
```

```python
# checking for null values
degree.isnull().sum()
```

```python
# checking for duplicates
duplicate = degree[degree.duplicated()]
duplicate
```

```python
# converting object to int data type
def convert(string):
    string = string.split(",")
    string = "".join(string)
    return int(string)
columns = degree.columns
columns = columns[3:]
for i in columns:
    degree[i] = degree[i].apply(convert)
```

```python
degree.head()
```

```python
i=degree[degree['Age Group']=='25 and older'].index
```

```python
degree.drop(i,inplace=True)
```

```python
# removing rows having sex='Total'
j=degree[degree['Sex']=='Total'].index
degree.drop(j,inplace=True)
```

```python
degree.head()
```

```python
plt.style.use('dark_background')
```

## EDA

```python
# total number of states
degree['State'].nunique()
```

```python
plt.figure(figsize=(10,5))
sns.pointplot(x='Sex',y='Bachelor\'s Degree Holders',data=degree,color='green',linestyles=['--'])
```

#### > More females are Bachelor's degree holders than males

### Which age group among the females have more bachelor's degree?

```python
df=degree[degree['Sex']=='Female']
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Age Group',y='Bachelor\'s Degree Holders',data=df,color='grey')
```

#### > Age group 40 to 64 has more number of bachelor's degree 

### So, in which particular field does the females of age group 40-64 are more bachelor's degree holders?

```python
df1=df[df['Age Group']=='40 to 64']
df1=df1[['Science and Engineering','Science and Engineering Related Fields','Business','Education','Arts, Humanities and Others']]
df1.head()
```

```python
plt.figure(figsize=(12,5))
sns.kdeplot(data=df1,palette='rainbow')
```

#### > Science and Engineering related fields have more bachelor degree holders 

### Which state has most number of bachelor degree holders?

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.scatterplot(x='State',y='Bachelor\'s Degree Holders',data=degree,color='red')
```

#### > California stands at first and Texas at second in number of bachelor's degree holders

### Which gender has the most bachelor degrees in California?

```python
df2=degree[degree['State']=='California']
df2.head()
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(x='Sex',y='Bachelor\'s Degree Holders',data=df2,palette='Set1')
```

#### > Above plot shows females have higher median than males thereby having more number of bachelor's degrees.

### So,which age group among the females in California has higher number of bachelor's degree?

```python
df3=df2[df2['Sex']=='Female']
df3.head()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x='Age Group',y='Bachelor\'s Degree Holders',data=df3,color='aqua')
```

#### > Age group 40 to 64 has more number of bachelor's degree

### So, in which particular field does the females of age group 40-64 are more bachelor's degree holders in California?

```python
df4=df3[df3['Age Group']=='40 to 64']
df4=df4[['Science and Engineering','Science and Engineering Related Fields','Business','Education','Arts, Humanities and Others']]
df4.head()
```

#### > Science and Engineering have more bachelor degree holders

```python
df5=degree[['Science and Engineering','Science and Engineering Related Fields','Business','Education','Arts, Humanities and Others']]
df5.head()
```

```python
col=df5.columns
col
```

```python
plt.figure(figsize=(12,8))
i=1
for j in col:
    plt.subplot(3,2,i)
    sns.barplot(x="Age Group", y=j, hue="Sex", data=degree,palette='magma')
    plt.title(j)
    i+=1
plt.subplots_adjust(wspace=0.6, hspace=0.6)
plt.show()
plt.tight_layout()
```

#### > Age group 40-64 has highest number of degree holders.
#### > Males have more bachelor's degree than females in 'Business' and 'Science and Engineering'.
#### > Females have more bachelor's degree than males in 'Education','Science and Engineering related fields' and 'Arts, Humanities and Others'.
