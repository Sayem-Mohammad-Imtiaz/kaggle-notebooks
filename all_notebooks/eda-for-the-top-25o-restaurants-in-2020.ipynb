```python
#Importing the required libraries

import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from pandas_profiling import ProfileReport
```

<h1 style="background-color:GREEN;font-family:Georgia;font-size:150%;text-align:center">1) DATA LOADING AND DATA PREPARATION </h1>

```python
top_250=pd.read_csv("/kaggle/input/restaurant-business-rankings-2020/Top250.csv")
```

```python
top_250.head()
```

```python
cap=top_250[top_250['YOY_Sales']==-21.2]
cap
```

```python
top_250.info()
```

Headquaters and content have many null values hence we can drop the columns

```python
#dropping of content and headquarters columns
top_250 = top_250.drop('Content', axis=1)
top_250 = top_250.drop('Headquarters', axis=1)
```

```python
top_250.shape
```

<h1 style="background-color:GREEN;font-family:Georgia;font-size:150%;text-align:center">2) Data Visualization</h1>

<h1 style="background-color:Orange;font-family:Georgia;font-size:150%;text-align:center"-->  Lets Analyse with respect to the features</h1>

  <h1 style="background-color:Purple;font-family:Georgia;font-size:150%;text-align:center">1) Sales </h1>

  <h1 style="background-color:Skyblue;font-family:Georgia;font-size:150%;text-align:left">1) Sales and Segment_Category </h1>

```python
grouped = top_250.groupby('Segment_Category').sum().reset_index()
grouped = grouped.sort_values('Sales', ascending=False).reset_index()
```

```python
fig = px.pie(grouped.head(15), values='Sales', names='Segment_Category', color_discrete_sequence=px.colors.sequential.RdBu)
fig.update_layout(title=' Sales Based on the Segment_Category  - Top 15')
```

<h1 style="background-color:GREEN;font-family:Georgia;font-size:100%;text-align:center">Splitting the Segment_Category into the Category </h1>

```python
splitcategory = top_250["Segment_Category"].str.split("&", n = -1, expand = True)
```

```python
top_250['category']=splitcategory[0]
```

```python
 top_250['category'].unique()
```

```python
grouped1 = top_250.groupby('category').sum().reset_index()
grouped1 = grouped1.sort_values('Sales', ascending=False).reset_index()
```

```python
fig = px.pie(grouped1.head(15), values='Sales', names='category', color_discrete_sequence=px.colors.sequential.RdBu)
fig.update_layout(title=' Sales Based on the top_15_Category ')
```

<h1 style="background-color:Skyblue;font-family:Georgia;font-size:175%;text-align:center"-->It clearly shows that the Quick Service type is most popular</scan>

```python
px.bar(top_250.category.value_counts(), color_discrete_sequence=px.colors.sequential.RdBu)
```

<h1 style="background-color:Skyblue;font-family:Georgia;font-size:150%;text-align:left">2) Sales and Resturant </h1>

```python
groupedres = top_250.groupby('Restaurant').sum().reset_index()
groupedres = groupedres.sort_values('Sales', ascending=False).reset_index()
fig = px.pie(groupedres.head(15), values='Sales', names='Restaurant', color_discrete_sequence=px.colors.sequential.RdBu)
fig.update_layout(title=' Sales Based on the Restaurant  - Top 15')
fig.update_traces(textposition='inside', textinfo='value')
```

  <h1 style="background-color:Red;font-family:Georgia;font-size:150%;text-align:center">McD and Starbucks are leading the Sales </h1>

  <h1 style="background-color:Purple;font-family:Georgia;font-size:150%;text-align:center">2) YOY_SALES </h1>

<h1 style="background-color:Skyblue;font-family:Georgia;font-size:150%;text-align:left">1) YOY_Sales and Segment_Category </h1>

```python
top_250['YOY_Sales']=top_250['YOY_Sales'].str.strip('%')
top_250['YOY_Sales']=top_250['YOY_Sales'].astype(float)
```

```python
px.bar(top_250.YOY_Sales.value_counts())
```

```python
groupedyoys = top_250.groupby('Segment_Category').sum().reset_index()
groupedyoys = groupedyoys.sort_values('YOY_Sales', ascending=False).reset_index()
fig = px.pie(groupedyoys.head(15), values='YOY_Sales', names='Segment_Category',template="seaborn")
fig.update_layout(title=' Year over Year increase in  Sales Based on the Segment_Category - Top 15')
fig.update_traces(textposition='inside', textinfo='value')
```

  <h1 style="background-color:Red;font-family:Georgia;font-size:150%;text-align:center">Fast Casual & Burger  is more popular in 2020 from  2019 </h1>

<h1 style="background-color:Skyblue;font-family:Georgia;font-size:150%;text-align:left">1) YOY_Sales and Restaurants </h1>

```python
yoy_sales_postive=top_250[top_250['YOY_Sales']>0]
yoy_sales_negative=top_250[top_250['YOY_Sales']<0]
```

```python
groupedyoyres = yoy_sales_postive.groupby('Restaurant').sum().reset_index()
groupedyoyres = groupedyoyres.sort_values('YOY_Sales', ascending=False).reset_index()
fig = px.pie(groupedyoyres.head(15), values='YOY_Sales', names='Restaurant',template="seaborn")
fig.update_traces(textposition='inside', textinfo='value')
fig.update_layout(title=' Year over Year increase in Sales Based on the Restaurant - Top 15')
```

  <h1 style="background-color:Skyblue;font-family:Georgia;font-size:150%;text-align:center">Chicken Salad Chick  is more popular in 2020 from  2019 </h1>

```python
groupedyoyresd = yoy_sales_negative.groupby('Restaurant').sum().reset_index()
groupedyoyresd = groupedyoyresd.sort_values('YOY_Sales', ascending=True).reset_index()
fig = px.bar(groupedyoyresd.head(15), x='Restaurant', y='YOY_Sales',
             hover_data=['Restaurant', 'YOY_Sales'], color='YOY_Sales',
             height=400)
fig.update_layout(title=' Year over Year Decrease in Sales Based on the Restaurant - Bottom 15')

fig.show()
```

  <h1 style="background-color:Skyblue;font-family:Georgia;font-size:150%;text-align:center">Houlihan's is having the highest decrease in Sales in  2020 from  2019 </h1>

 <h1 style="background-color:Purple;font-family:Georgia;font-size:150%;text-align:center">3) YOY_UNITS </h1>


<h1 style="background-color:Skyblue;font-family:Georgia;font-size:150%;text-align:left">1) YOY_UNITS and Segment_Category </h1>

```python
top_250['YOY_Units']=top_250['YOY_Units'].str.strip('%')
top_250['YOY_Units']=top_250['YOY_Units'].astype(float)
```

```python
px.bar(top_250.YOY_Units.value_counts())
```

  <h1 style="background-color:Blue;font-family:Georgia;font-size:150%;text-align:center">This shows that 24 restaurants have retained there YOY_units and no increase or decrease in yoy_units </h1>

```python
yoy_units_postive=top_250[top_250['YOY_Units']>0]
yoy_units_negative=top_250[top_250['YOY_Units']<0]
```

```python
groupedyoyu = top_250.groupby('Segment_Category').sum().reset_index()
groupedyoyu = groupedyoyu.sort_values('YOY_Units', ascending=False).reset_index()
fig = px.pie(groupedyoyu.head(15), values='YOY_Units', names='Segment_Category',template="seaborn")
fig.update_layout(title=' Year over Year increase in Units  Based on the Segment_Category - Top 15')
fig.update_traces(textposition='inside', textinfo='value')
```

  <h1 style="background-color:Red;font-family:Georgia;font-size:150%;text-align:center">Fast Casual & Burger  is having highest increase in Units in 2020 from  2019 </h1>

```python
groupedyoyresu = yoy_units_postive.groupby('Restaurant').sum().reset_index()
groupedyoyresu = groupedyoyresu.sort_values('YOY_Units', ascending=False).reset_index()
fig = px.pie(groupedyoyresu.head(15), values='YOY_Units', names='Restaurant',template="seaborn")
fig.update_layout(title=' Year over Year increase in Units Based on the Restaurant - Top 15')
fig.update_traces(textposition='inside', textinfo='value')
```

  <h1 style="background-color:Red;font-family:Georgia;font-size:150%;text-align:center"> Chicken Salad Chick is more popular in 2020 from  2019 </h1>

```python
groupedyoyresd1 = yoy_units_negative.groupby('Restaurant').sum().reset_index()
groupedyoyresd1 = groupedyoyresd1.sort_values('YOY_Units', ascending=True).reset_index()
fig = px.bar(groupedyoyresd1.head(15), x='Restaurant', y='YOY_Units',
             hover_data=['Restaurant', 'YOY_Units'], color='YOY_Units',
             height=400)
fig.update_layout(title=' Year over Year Decrease in Units Based on the Restaurant - Bottom 15')

fig.show()
```

 <h1 style="background-color:Skyblue;font-family:Georgia;font-size:150%;text-align:center">Houlihan's is having the highest decrease in Units in  2020 from  2019 </h1>
