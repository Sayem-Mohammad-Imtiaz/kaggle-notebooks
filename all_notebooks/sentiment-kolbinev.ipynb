Проводим синтаксический анализ текста

Первый пример

```python
# Определяем синтаксические правила

# S -> формат предложений
# P - preposition (предлог)
# N - noun (существительное)
# V - verb (глагол)
# Det - determination (определение)

syntex_rules = nltk.CFG.fromstring("""
S -> NP VP
PP -> P NP
NP -> Det N | Det N PP | 'лингвистика' | N Det N
VP -> V NP | VP PP | V N N Det N
Det -> 'своей' | 'искусственного' | 'Математическая' | Det N
N -> 'лингвистика' | 'ветвью' | 'платформе' | 'науки' | 'интеллекта' | Det N
V -> 'является'
P -> 'на'
""")

# Задаем предложение для разбора
sentence = ['Математическая', 'лингвистика', 'является', 'ветвью', 'науки', 'искусственного', 'интеллекта']

# Разбираем с помощью ранее определенных правил предложение
parser = nltk.ChartParser(syntex_rules)

# Выводим результат
print(list(parser.parse(sentence))[0])
```

```python
# Определяем синтаксические правила

# S -> формат предложений
# P - preposition (предлог)
# N - noun (существительное)
# V - verb (глагол)
# Det - determination (определение)

syntex_rules = nltk.CFG.fromstring("""
S -> NP VP
PP -> P NP
NP -> Det N | Det N PP | 'лингвистика' | N Det N
VP -> V NP | VP PP | V N N Det N | V N
VN -> V N
Det -> 'Главным' | 'искусственного' | 'Математическая' | Det N
N -> 'организатором' | 'ветвью' | 'платформе' | 'науки' | 'Н. Д. Андреев' | Det N | 'Совещания'
V -> 'был'
P -> 'на'
""")

# Задаем предложение для разбора
sentence = ['Главным', 'организатором', 'Совещания', 'был', 'Н. Д. Андреев']

# Разбираем с помощью ранее определенных правил предложение
parser = nltk.ChartParser(syntex_rules)

# Выводим результат
print(list(parser.parse(sentence))[0])
```
