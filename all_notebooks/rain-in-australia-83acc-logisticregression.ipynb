# Data Preprocessing

```python
# Missing values in percentages
df.isna().sum()/df.shape[0]*100
```

```python
# Dropping 'DATE' and other columns that have more than 30% of missing data
df = df.drop(['Date','Evaporation', 'Sunshine', 'Cloud9am', 'Cloud3pm'], axis=1)
df
```

```python
df.info()
```

```python
# Separating the categorical type and numerical  columns
cat_cols = df.select_dtypes(include = ['object'])
num_cols = df.select_dtypes(exclude = ['object'])
```

```python
for cols in cat_cols:
    cat_cols.fillna(method='ffill', inplace=True )

for col in num_cols:
    num_cols[col].fillna(num_cols[col].mean(), inplace=True)
```

```python
num_cols.info()
```

```python
cat_cols.info()
```

```python
# lable encoding the category columns with label Encoder
lb_enc = LabelEncoder()
cat_cols = cat_cols.astype(str).apply(lb_enc.fit_transform)
```

```python
# Joining both 
df = pd.concat([cat_cols, num_cols], axis = 1)
```

```python
df.info()
```

# Data Modelling

```python
X = df.drop('RainTomorrow', axis=1)
y = df['RainTomorrow']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
np.random.seed(42)
model = LogisticRegression(solver='liblinear',
                           multi_class='ovr')
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_preds = model.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_preds))
```

```python
print(classification_report(y_test, y_preds))
```
