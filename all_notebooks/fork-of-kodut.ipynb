```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

%matplotlib inline
pd.set_option('display.max_rows', 20)

from subprocess import check_output
print(check_output(["ls", "../input"]).decode("utf8"))
```

```python
df = pd.read_csv("../input/vgsaleeeeee/vgsales.csv")
df
```

Selles andmestikus on videomängud, mida on müüdud rohkem, kui 100 000 koopiat. Andmestik on jaotatud 11-ks tulbaks, mis iseloomustavad mängu ja selle positsiooni teiste suhtes. 
Meie uurimisküsimuseks on videomängude läbimüük uuritavate aastate jooksul nii üldiselt, kui ka kolmes peamises piirkonnas: Euroopa, Jaapan, Põhja-Ameerika.

Siin on näha, millised on läbi aegade enim mänge müünud konsoolid.

```python
fig,ax = plt.subplots(figsize=(10,3))
df["Platform"].value_counts(sort=True).plot(kind="bar",ax=ax,rot=90)
plt.xlabel("Platform")
plt.ylabel("Number of sales")
```

Histogrammilt on näha, et kõige suuremate mängude müüginumbritega konsool on Nintendo DS ja väga lähedal kannul on PlayStation 2.

Nende tabelite pealt tuleb välja, milline seos on globaalsel läbimüügil erinevate regioonidega.

```python
df.plot.scatter("Global_Sales","EU_Sales",alpha = 0.1)
df.plot.scatter("Global_Sales","NA_Sales",alpha = 0.1)
df.plot.scatter("Global_Sales","JP_Sales",alpha = 0.1)
```

Tuleb välja, et kõige vähem globaalsetest trendidest mõjutatud regioon on Jaapan ja kõige rohkem Põhja-Ameerika.

Järgnevalt graafikult vaatame, millise firma mängud olid kõige populaarsemad.

```python
f,ax=plt.subplots(1,1,figsize=(12,5))
df.groupby("Publisher")["Global_Sales"].sum().sort_values(ascending=False)[:10].plot(kind="bar")
ax.set_ylabel("Kasum")
ax.set_xlabel("Tootja")
ticks=plt.setp(ax.get_xticklabels(), rotation=20, fontsize=7)
```

Tuleb välja, et kõige populaarsem mängude tootja on suure ülekaaluga Nintendo.
