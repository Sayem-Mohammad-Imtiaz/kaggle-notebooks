```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf
import os.path as path
from PIL import Image as image
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../input/plastic-dataset/labels.csv")
df.head()
```

```python
def expand_path(p):
    if path.isfile('../input/plastic-dataset/images/' + p + '.jpeg'): return '../input/plastic-dataset/images/' + p + '.jpeg'
    return p 
df['image_path'] = df['Filenames']
df['image_path'] = df['image_path'].apply(expand_path)
```

```python
def process(x):
    imageP = image.open(x)
    width, height = imageP.size
    if (width > 3025):
        return np.asarray(imageP.resize((341,256 )).rotate(270, expand=True))
    else:
        return np.asarray(imageP.resize((256, 341)))
    
df['images'] = df['image_path'].map(process)
df.to_pickle('./resized-in-df.pkl')
```
