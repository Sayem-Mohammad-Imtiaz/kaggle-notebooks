```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
import graphviz
import sklearn
```

```python
train = pd.read_csv("../input/titanic/train.csv").drop(columns=["PassengerId", "Name", "Ticket", "Cabin"])
test = pd.read_csv("../input/titanic/test.csv").drop(columns=["PassengerId", "Name", "Ticket", "Cabin"])

train = train.replace({"Embarked": {"Q":1, "S":2, "C":3}, "Sex": {"male":0, "female":1}, "Survived": {0:"Dead",1:"Survive"}})
test = test.replace({"Embarked": {"Q":1, "S":2, "C":3}, "Sex": {"male":0, "female":1}, "Survived": {0:"Dead",1:"Survive"}})

train = train.dropna()
test = test.dropna()
```

```python
train.head(5)
```

```python
train_data = train.iloc[:, lambda col: [1,2,3,4,5,6,7]]
train_target = train.iloc[:,0].values
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(train_data, train_target)
accuracies = cross_val_score(estimator = classifier, X = train_data, y = train_target, cv = 10)
print("Accuracy: {:.2f} %".format(accuracies.mean()*100))
print("Standard Deviation: {:.2f} %".format(accuracies.std()*100))
```

```python
tree = DecisionTreeClassifier().fit(train_data, train_target)
dot_data = sklearn.tree.export_graphviz(tree, out_file = None, feature_names = train_data.columns)
g1 = graphviz.Source(dot_data)
g1
```
