# ****Reading DataSet:-****

```python
df = pd.read_csv('../input/building-permit-applications-data/Building_Permits.csv')
df.head()
```

# **Dividing DataSet in to Location Details with Target Variable (Site Permit) :-**

```python
df_loc = df[['Block','Lot','Street Number','Street Number Suffix','Street Name','Street Suffix','Unit','Unit Suffix','Neighborhoods - Analysis Boundaries','Location','Zipcode','Site Permit']]
df_loc.head()
```

```python
df_loc.info()
```

# Finding percentage of Null Values:-****

```python
(df_loc.isnull().sum()/len(df)*100).sort_values(ascending=False)
```

We drop Columns 'Street Number Suffix','Unit','Unit Suffix' as it has more than 95% of NULL Values

We Use Use BackWard Fill Method to Impute the Null Values.

***a) Treating Target Variable:-***

```python
print(df['Site Permit'].unique())
#Y = Yes
#nan = No <-- we replace nan here with 'N'
```

```python
df_loc['Site Permit'] = df_loc['Site Permit'].replace(to_replace=np.nan , value='N')
```

```python
df_loc['Site Permit'].unique()
```

***b) Missing value imputation :-***

```python
# filling Null Values
df_loc_null = df_loc.fillna(method='bfill',axis = 0).fillna(0)
```

```python
df_loc_null.isnull().sum()
```

# Outlier Detection:-

```python
df_num = df_loc_null._get_numeric_data()
df_num.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for i in df_num:
    plt.figure(figsize=(10,5))
    sns.boxplot(x=df_num[i])
    plt.show()
    df_num[i].skew()
```
