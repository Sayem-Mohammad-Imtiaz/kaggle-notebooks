```python
#Descobrindo a quantidade de pessoas mortas

plt.figure(figsize=(45,10))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9,
                      wspace=0.5, hspace=0.2)
plt.subplot(141)
plt.title('death_yn ',fontsize = 40)
dados['death_yn'].value_counts().plot.pie(autopct="%1.1f%%")
```

* 2,2% = Mortos
* 12,4% = Não-Identificados
* 41,5% = Ausentes/Perdidos
* 43,9% = Vivos

```python
plt.figure(figsize=(10,8))
sb.countplot(y="age_group",hue='current_status', data=dados)
```

>É possivel identificar que casos na faixa de 10-19 anos é menor que a de idosos de 60-69 anos mesmo com todas medidas de preucaução e incentivos a manter a isolação.
