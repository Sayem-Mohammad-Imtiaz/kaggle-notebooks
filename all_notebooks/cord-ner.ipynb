**Release of CORD-NER Data: Automated Recognition and Typing of Named Entities of CORD-19 Data**

In response to the Call to Action to the Tech Community on [New Machine Readable COVID-19 Dataset](https://www.cccblog.org/2020/03/16/news-call-to-action-to-the-tech-community-on-new-machine-readable-covid-19-dataset/?utm_source=feedblitz&utm_medium=FeedBlitzRss&utm_campaign=cccblog) from White House on March 16, researchers from the Allen Institute for AI, Chan Zuckerberg Initiative (CZI), Georgetown Universityâ€™s Center for Security and Emerging Technology (CSET), Microsoft, and the National Library of Medicine (NLM) at NIH (the National Institutes of Health) released CORD-19, the [COVID-19 Open Research Dataset](https://www.kaggle.com/allen-institute-for-ai/CORD-19-research-challenge) of scholarly literature about COVID-19, SARS-CoV-2, and the coronavirus group.

To facilitate biomedical research on the CORD-19 dataset, our Data Mining Group at Computer Science, University of Illinois, has been developing an automated, comprehensive named entity annotation and typing system, CORD-NER, and has generated automatically annotated CORD-NER dataset, based on the CORD-19 dataset. This automatically annotated dataset has been published at the ArXiv website: [https://arxiv.org/abs/2003.12218](https://arxiv.org/abs/2003.12218) and will be updated timely based on the incremental updates of the CORD-19 dataset and the improvement of our system.

The distinctive features of this published dataset include (1) it covers 75 fine-grained entity types: In addition to the common biomedical entity types (e.g., genes, chemicals and diseases), it covers many new entity types specifically related to the COVID-19 studies (e.g., coronaviruses, viral proteins, evolution, materials, substrates and immune responses), which may benefit research on COVID-19 related virus, spreading mechanisms, and potential vaccines; (2) it relies on distantly- and weakly-supervised NER methods, with no need of expensive human annotation on any articles or subcorpus; and (3) its entity annotation quality surpasses SciSpacy (over 10% higher on the F1 score based on a sample set of documents), a fully supervised BioNER tool. SciSpacy requires human effort for training data annotation and covers only 18 types. Our NER system supports incrementally adding new documents as well as adding new entity types when needed by adding dozens of seeds as the input examples.

We are in the process of improving our CORD-NER system and developing a set of new text mining tools on this dataset. The dataset has been used to feed to a CORD-19 knowledge graph construction engine and a natural language question-answering engine (both led by Professor Heng Ji at UIUC) and to a SetSearch engine (led by Jiaming Shen at UIUC).
