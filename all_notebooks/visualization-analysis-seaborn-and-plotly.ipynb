## 3-Bar Plot

-sns.barplot() ---> Selection of graph type.

-x and y ---> Series

-figsize() ---> Takes 2 input for dimensions of graph.(x and y)

-xticks(rotation=) ---> Takes input as an angle to rotate x axis indexes. 


<a id='4'></a>
### 3.1-Which game platforms were most preferred?

  This bar plot (Platforms) was created to show amount of platforms selection for games. As it can be seen in the graph, DS and PS2 are most selected platform. As most of datas belongs to past years, this visualization can be acceptable.


```python
data['Values']=1
plt.figure(figsize=(10,10))

sns.barplot(x=data.groupby('Platform').sum().sort_values('Values',ascending=False).index,
            y=data.groupby('Platform').sum().sort_values('Values',ascending=False).Values
           )
plt.title('Numbers of Platforms Preferred')
plt.xticks(rotation=45)
plt.show()
```

<a id='5'></a>
### 3.2-Which genres were the most preferred?
Between 1980 and 2020, the most preferred game type are action games.


```python
plt.figure(figsize=(12,10))
sns.barplot(x=data.groupby('Genre').sum().sort_values('Values',ascending=False).index,
            y=data.groupby('Genre').sum().sort_values('Values',ascending=False).Values)
plt.title('Preferred Game Types')
plt.show()
```

<a id='6'></a>
### 3.3-Most Purchased Game Types
Action games are both the most preferred game type and the most paid game type totally.


```python
plt.figure(figsize=(12,10))
sns.barplot(x=data.groupby('Genre').sum().sort_values('Global_Sales',ascending=True).index,
            y=data.groupby('Genre').sum().sort_values('Global_Sales',ascending=True).Global_Sales)
plt.title('Most purchased game types ')
plt.show()
```

<a id='7'></a>
### 3.4-Numbers of games for publishers
The platform that produces the most games is Eletronics Arts.

```python
plt.figure(figsize=(10,10))
new_Data=data.groupby('Publisher').sum().sort_values('Values',ascending=False)
new_Data[new_Data.Values>100]
sns.barplot(x=new_Data[new_Data.Values>100].Values,
            y=new_Data[new_Data.Values>100].index)
plt.xticks(rotation=0)
plt.title('Publishers vs Game Numbers (greater than 100) ')
plt.show()
```

## 4-Point Plot

-sns.pointplot() ---> Selection of graph type.

-x and y ---> Series

-xlabel and ylabel ---> Titles of axis 

-text ---> It provides writing on the graph. Location,type and color can be determined.

-grid() ---> Shows the lines in the background.

-xticks(rotation=) ---> Takes input as an angle to rotate x axis indexes. 

<a id='9'></a>
### 4.1-Number of sales of games in regions



```python
plt.subplots(figsize=(17,10))
sns.pointplot(x=data.groupby('Year').sum().index,
              y=(data.groupby('Year').sum().EU_Sales))
sns.pointplot(x=data.groupby('Year').sum().index,
              y=(data.groupby('Year').sum().NA_Sales),color='red')
sns.pointplot(x=data.groupby('Year').sum().index,
              y=(data.groupby('Year').sum().JP_Sales),color='green')

plt.xticks(rotation=45)
plt.ylabel('Sales')
plt.xlabel('Years')
plt.text(34,330,'Europe',color='blue',fontsize=15)
plt.text(34,315,'North America',color='red',fontsize=15)
plt.text(34,300,'Japan',color='green',fontsize=15)
plt.title('Sales in Regions by Years')
plt.grid()
plt.show()
```

<a id='10'></a>
### 4.2-Using of PC changing

```python
plt.figure(figsize=(15,10))
sns.pointplot(x=data[data.Platform=='PC'].groupby('Year').sum().sort_values('Year',ascending=True).index,
              y=data[data.Platform=='PC'].groupby('Year').sum().sort_values('Year',ascending=True).Values)
plt.xticks(rotation=45)
plt.title('PC Preference by Years ')
plt.xlabel('Years')
plt.grid()
plt.show()
```

## 5-Pie Plot

-plt.pie ---> Selection of graph type.

-explode ---> Distance between slices.

-autopct ---> It shows values on the graph and digit number of values can be determined.

-labels ---> Titles of slices.

<a id='12'></a>
### 5.1-The Ratios of Regions in Global Sales
It can easily be seen that the region with the highest rate of total sales is the EU, and this rate is approximately 50%.


```python
plt.figure(figsize=(8,8))
sizes=[data.EU_Sales.sum(),data.NA_Sales.sum(),data.JP_Sales.sum(),data.Other_Sales.sum()]
plt.pie(sizes,labels=('North America','Europe','Japan','Other'),explode=(0.1,0.1,0.0,0.0),autopct='%1.1f%%')
plt.title('Regions Share in Sales',size=15)
plt.show()
```

## 6-Heat Map

-corr() ---> Makes correlation between given data.

-linewidths ---> Space between small regions.

-annot ---> Shows coleration variables.

<a id='14'></a>
### 6.1-What is the correlation between sales region?
Correlations of sales volumes in the regions with each other can be seen in the map chart. As the correlation ratio approaches 1, the positive linear relationship increases. The most positive relationship between EU_Sales and Global can be seen in this chart. The proportion in the pie chart can prove this.


```python
plt.figure(figsize=(8,8))
sns.heatmap(data[['EU_Sales','NA_Sales','JP_Sales','Other_Sales','Global_Sales']].corr(),linewidths=0.5,annot=True)
plt.title('Correlation Map Between Sales',size=15)
plt.show()
```

<a id='15'></a>
### 6.2-In which region, which genres were the most sold ?
The interest in action games is quite high in the North America region.

```python
plt.figure(figsize=(15,10))
sns.heatmap(data[['Genre','EU_Sales','NA_Sales','JP_Sales','Other_Sales']].groupby('Genre').sum(),annot=True,fmt = '.1f',linewidth=0.5)
sns.set(font_scale=1)
plt.title('Most Prefered Genre by Region',size=15)
plt.show()
```

## 7-3D Scatter

-x,y and z ---> Series

-margin(l,r,b,t) ---> Left,right,bottom,top

<a id='17'></a>
### 7.1-EU,JP and NA Sales on 3D

```python
trace1=go.Scatter3d(x=data.EU_Sales,
                    y=data.JP_Sales,
                    z=data.NA_Sales,
                    mode='markers',
                    marker=dict(size=10))
data2=[trace1]
layout=dict(margin=dict(l=0,r=0,b=0,t=0))
fig=dict(data=data2,layout=layout)
iplot(fig)
```

<a id='18'></a>
## 8-Line Plot

-go.Scatter ---> Type of graph.

-mode ---> type of plot like marker.

-color ---> color of lines. RGB (red, green, blue)

-layout ---> It allows adding features for graphics by using it in dictionary type. Title, length of x axis ticks, zero line etc.

<a id='19'></a>
### 8.1-In which year were the most games sold?
The most game sales were made in 2008 according to global values.


```python
trace1=go.Scatter(x=data.groupby('Year').sum().sort_values('Year',ascending=True).index,
                  y=data.groupby('Year').sum().sort_values('Year',ascending=True).Global_Sales,
                  mode='lines',
                  marker=dict(color='rgba(180,1,1,0.8)'))
data4=[trace1]
layout=dict(title='Global Sales by Years',
            xaxis=dict(title='Global Sales'),
            yaxis=dict(title='Years'))
fig=dict(data=data4,layout=layout)
iplot(fig)
```

## 10-Scatter Plot

```python
df1=data[data.Year==2007]
df2=data[data.Year==2008]
df3=data[data.Year==2009]

trace1 =go.Scatter(
                    x = df1.Global_Sales,
                    y = df1.EU_Sales,
                    mode = "markers",
                    name = "2007",
                    marker = dict(color = 'rgba(255, 128, 255, 0.8)'),
                  )

trace2 =go.Scatter(
                    x = df2.Global_Sales,
                    y = df2.EU_Sales,
                    mode = "markers",
                    name = "2008",
                    marker = dict(color = 'rgba(255, 128, 2, 0.8)'),
                    )

trace3 =go.Scatter(
                    x = df3.Global_Sales,
                    y = df3.EU_Sales,
                    mode = "markers",
                    name = "2009",
                    marker = dict(color = 'rgba(0, 255, 200, 0.8)'),
                    )

data = [trace1, trace2, trace3]
layout = dict(title = 'EU sales and Global Sales Chart in the years of highest sales (200,2008,2009)',
              xaxis= dict(title= 'Global Sales',ticklen=7,zeroline= False),
              yaxis= dict(title= 'EU Sales',ticklen= 7,zeroline= False)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```
