#Check Musical Valence. How positivity of the music has changed over the last 70 years.

```python
#I wanted only the Decades years, though here is anually.

fig, axes = plt.subplots(1, 1, figsize=(14, 6))
sns.boxplot(x='release_date', y='valence', data=df, showfliers=False);
plt.xticks(rotation=45)
```

The median valence for the 2010s was the lowest measured at 0.46, meaning that the average track from the last decade was more likely to have a slight negative feeling than not.

```python
fig = px.box(df, x="release_date", y="valence", title='Track Valence through the Years')
fig.show()
```

```python
fig = px.bar(df, x= "release_date", y= "valence", color_discrete_sequence=['crimson'], title='Valence through the years')
fig.show()
```

#A track that scores higher for valence and lower for energy would indicate a more chilled-out, happy #track. Whereas the opposite (low valence, high energy) would be more of a frantic or angry sounding #song.
https://www.linkedin.com/pulse/change-musical-mood-1950-2019-kevin-stevens?articleId=6696344296271962113

```python
fig = px.scatter(df, x="energy", y="valence",color_discrete_sequence=['#27F1E7'],text='release_date', title='Scores of energy and Valence' )
fig.show()
```

```python
fig = go.Figure(data=[go.Scatter(
    x=df['energy'][0:10],
    y=df['valence'][0:10],
    mode='markers',
    marker=dict(
        color=[145, 140, 135, 130, 125, 120,115,110,105,100],
        size=[100, 90, 70, 60, 60, 60,50,50,40,35],
        showscale=True
        )
)])
fig.update_layout(
    title='Score of Energy and Valence',
    xaxis_title="energy",
    yaxis_title="Valence",
)
fig.show()
```

```python
px.scatter(df, x='energy', y='valence', color='release_date', log_y=True, marginal_y='box', range_x=[-20, 200])
```

```python
fig = px.bar(df[['release_date','genre']].sort_values('genre', ascending=False), 
                        y = "genre", x= "release_date", color='genre', template='ggplot2')
fig.update_xaxes(tickangle=45, tickfont=dict(family='Rockwell', color='crimson', size=14))
fig.update_layout(title_text="Genre through the Years")
fig.show()
```

```python
fig = px.bar(df[['release_date','topic']].sort_values('topic', ascending=False), 
                        y = "topic", x= "release_date", color='topic', template='ggplot2')
fig.update_xaxes(tickangle=45, tickfont=dict(family='Rockwell', color='crimson', size=14))
fig.update_layout(title_text="Topics through the Years")

fig.show()
```

#Danceability of a track – as Spotify calls it – has increased since the 1950s.

```python
fig = px.violin(df, y="danceability", x="release_date", color="genre", box=True, points="all",
          hover_data=df.columns)
fig.show()
```
