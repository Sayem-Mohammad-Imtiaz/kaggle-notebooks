```python
import pandas as pd
import numpy as np

data=pd.read_csv('../input/my-dataset/credit_train.csv')
data.shape
```

```python
data[-515:].head() ## data beyond 100000 is null !!! so we will remove it !!
```

```python
data=data[:-514]
```

```python
# Filling the mean of the clluster that we have found on grouping to most co-related attribute
data["Credit Score"]=data.groupby(["Current Loan Amount"])["Credit Score"].transform(lambda x: x.fillna(x.mean()))
data["Credit Score"]=data.groupby(["Years of Credit History"])["Credit Score"].transform(lambda x: x.fillna(x.mean()))
```

```python
# dealing outliers
data['Credit Score']=data['Credit Score'].apply(lambda x: x/10 if x>1000 else x)
```

```python
# Filling the mean of the clluster that we have found on grouping to most co-related attribute
data["Annual Income"]=data.groupby(["Number of Open Accounts"])["Annual Income"].transform(lambda x: x.fillna(x.mean()))
data["Annual Income"]=data.groupby("Years of Credit History")["Annual Income"].transform(lambda x: x.fillna(x.mean()))
```

```python
#replacing NAN with 0 for those person Who haven't Commited any kind of Crimes in past Months
data['Months since last delinquent']=data['Months since last delinquent'].fillna(0)
data.reset_index(drop=True, inplace=True)
```

```python
# Filling the mean of the clluster that we have found on grouping to most co-related attribute
data["Years in current job"]=data.groupby(["Monthly Debt"])["Years in current job"].transform(lambda x: x.fillna(x.mode()))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pylab
import scipy.stats as stats

def diagnostic_plots(df, variable):
    # function takes a dataframe (df) and
    # the variable of interest as arguments

    # define figure size
    plt.figure(figsize=(16, 4))

    # histogram
    plt.subplot(1, 3, 1)
    sns.distplot(df[variable], bins=40)
    plt.title('Histogram')

    # Q-Q plot
    plt.subplot(1, 3, 2)
    stats.probplot(df[variable], dist="norm", plot=pylab)
    plt.ylabel('RM quantiles')

    # boxplot
    plt.subplot(1, 3, 3)
    sns.boxplot(y=df[variable])
    plt.title('Boxplot')

    plt.show()
```
