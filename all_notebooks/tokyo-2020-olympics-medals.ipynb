```python
total_medals = {'Gold_Medal': df.Gold_Medal.sum(), 
                'Silver_Medal': df.Silver_Medal.sum(),
                'Bronze_Medal': df.Bronze_Medal.sum()}
```

```python
for i in total_medals.keys():
    tmp_df = df.copy()
    tmp_df[f'{i}'] =  tmp_df[f'{i}'].apply(lambda x: x / total_medals.get(f'{i}') * 100)
    tmp_df.rename(columns={f'{i}': f'{i}' + '_percent_of_total'}, inplace=True)
    tmp_df = tmp_df[[f'{i}'+ '_percent_of_total']]
    df = pd.concat([df, tmp_df], axis=1)
```

```python
df.head()
```

```python
df = df.round()
```

```python
tmp_team_medals = {'Gold_Medal': [], 
                'Silver_Medal': [],
                'Bronze_Medal': []}
for medal in total_medals.keys():
    df_team_medals_percent = pd.DataFrame()
    for i in range(len(df)):
        tmp_team_medals[medal].append(df[medal][i] / df['Total'][i] * 100)
    df_team_medals_percent[f'{medal}' + 'medal_in_the_team_event'] = tmp_team_medals[medal]
    df = pd.concat([df, df_team_medals_percent], axis=1)
```

```python
def medals_the_team(country):
    get_stats = df.loc[df['Country'] == country]
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))
    labels = ['Gold', 'Silver', 'Bronze']
    width = 0.4
    x = np.arange(len(labels))
    colors = ['#ffd700','#c0c0c0','#cd7f32']
    ax1.set_xticks(x)
    ax1.set_xticklabels(labels)
    ax2.set_xticks(x)
    ax2.set_xticklabels(labels)
    
    fig.suptitle(f'{country}, Total: {get_stats.iloc[0,4]}, Rank By Total:{get_stats.iloc[0,5]}', size=15)
    ax1.bar(get_stats.columns[1:4], get_stats.iloc[0,0:4][1:4], width=0.8, color = colors)
    ax2.bar(x - width/2, get_stats.iloc[0,6:9], width, label='percentage of total', color='tab:gray')
    ax2.bar(x + width/2, get_stats.iloc[0,9:12], width, label='percentage of team', color='tab:brown')
    ax2.legend(loc=3)
```

```python
medals_the_team('United States of America')
```

```python
medals_the_team('ROC')
```
