## Glass Type Classification  

Given *data about the composition of various glassses*, let's try to classify the **type** of a given glass.  
  
We will use PyCaret to train and compare 15 different models on our data.

# Getting Started

```python
!pip install pycaret
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from pycaret.classification import setup, models, compare_models, predict_model
```

```python
data = pd.read_csv('../input/glass/glass.csv')
```

```python
data
```

```python
data.info()
```

# Visualization

```python
corr = data.corr()

plt.figure(figsize=(16, 14))
sns.heatmap(corr, annot=True, vmin=-1.0, vmax=1.0, cmap='mako')
plt.title("Correlations")
plt.show()
```

```python
plt.figure(figsize=(12, 12))
plt.pie(data['Type'].value_counts(), labels=data['Type'].value_counts().index, colors=sns.color_palette('mako'))
plt.title("Class Distribution")
plt.show()
```

# Model Comparison and Evaluation

```python
setup(
    data=data,
    target='Type',
    normalize=True,
    train_size=0.7
)
```

```python
models()
```

```python
best_model = compare_models()
```

```python
predict_model(best_model)
```
