**The number of Non fiction books is more than Fiction.**

```python
#User rating 
g = sns.FacetGrid(df,col = 'Genre')
g.map(sns.distplot,'User Rating',bins = 25)
g.add_legend()
```

```python
#Price
g = sns.FacetGrid(df,col = 'Genre')
g.map(sns.distplot,'Price',bins = 25)
g.add_legend()
```

```python
#Reviews
g = sns.FacetGrid(df,col = 'Genre')
g.map(sns.distplot,'Reviews',bins = 25)
plt.show()
```

```python
df1 = df[df['Genre']=='Fiction']
df2 = df[df['Genre']=='Non Fiction']

rdf1 = df1['User Rating'].value_counts().reset_index()
rdf2 = df2['User Rating'].value_counts().reset_index()

tr1 = go.Bar(x = rdf1['index'],
             y = rdf1['User Rating'],
             name = 'Fiction',
             marker = dict(color = 'LightSeaGreen')
            
            )

tr2 = go.Bar(x = rdf2['index'],
             y = rdf2['User Rating'],
             name = 'Non Fiction',
             marker = dict(color = 'RoyalBlue')
            
            )

data = [tr1,tr2]
layout = go.Layout(title = "Fiction & Non Fiction",
                   title_x = 0.5,
                   title_font = dict(size = 16, color = 'RoyalBlue'),
                   xaxis = dict(title = 'Genre'),
                   yaxis = dict(title = 'User Rating'),
                   template = 'plotly_white'
                  
                  )
fig = go.Figure(data = data, layout = layout)
fig.show()
```

* **Fiction books - The value(4.8) of user rating is the most than others.**
* **Non Fiction books - The values(4.6 and 4.7) are the most than others.**

```python
df1 = df[df['Genre']=='Fiction']
df2 = df[df['Genre']=='Non Fiction']

rdf1 = df1['Year'].value_counts().reset_index()
rdf2 = df2['Year'].value_counts().reset_index()

tr1 = go.Bar(x = rdf1['index'],
             y = rdf1['Year'],
             name = 'Fiction',
             marker = dict(color = 'LightSeaGreen')
            
            )

tr2 = go.Bar(x = rdf2['index'],
             y = rdf2['Year'],
             name = 'Non Fiction',
             marker = dict(color = 'RoyalBlue')
            
            )

data = [tr1,tr2]
layout = go.Layout(title = 'The number of Non Fiction and Fiction books by years',
                   title_x = 0.5,
                   title_font = dict(size = 16, color = 'RoyalBlue'),
                   xaxis = dict(title = 'Year'),
                   yaxis = dict(title = 'Count'),
                   template = 'plotly_white'
                  
                  )

fig = go.Figure(data = data, layout = layout)
fig.show()
```

* **Fiction books is significantly increased in 2014 and then sharply decreased in 2015.**
* **Non fiction books is significantly descreased in 2014, and then sharply increased in 2015.**
* **The number of Non fiction books is more than Fiction books every year, except 2014.**
* **Both Fiction and Non fiction books are inconsistent on their counts by years.**

```python
df1 = df[df['Genre']=='Fiction']
df2 = df[df['Genre']=='Non Fiction']

rdf1 = df1.groupby("Year").mean().reset_index()
rdf2 = df2.groupby("Year").mean().reset_index()

tr1 = go.Scatter(x = rdf1['Year'],
                 y = rdf1['User Rating'],
                 name = 'Fiction',
                 mode = 'lines+markers',
                 marker = dict(color = 'LightSeaGreen',line = dict(color = 'AliceBlue',width = 1.5))
                
                )

tr2 = go.Scatter(x = rdf2['Year'],
                 y = rdf2['User Rating'],
                 name = 'Non Fiction',
                 mode = 'lines+markers',
                 marker = dict(color = 'RoyalBlue', line = dict(color = 'AliceBlue',width = 1.5))
                
            )
data = [tr1,tr2]
layout = go.Layout(title = 'Average User Rating over the years',
                   title_x = 0.5,
                   title_font = dict(size = 16, color = 'RoyalBlue'),
                   xaxis = dict(title = 'Year'),
                   yaxis = dict(title = 'User Rating'),
                   template = 'plotly_white'
                  
                  )

fig = go.Figure(data = data, layout = layout)
fig.show()
```

* **Fiction books is sharply descreased in 2012, but then sharply increased from 2012.**
* **Non fiction books is descreased in 2017, but then increased from 2018.**


```python
df1 = df[df['Genre']=='Fiction']
df2 = df[df['Genre']=='Non Fiction']

rdf1 = df1.groupby("Year").mean().reset_index()
rdf2 = df2.groupby("Year").mean().reset_index()

tr1 = go.Scatter(x = rdf1['Year'],
                 y = rdf1['Reviews'],
                 name = 'Fiction',
                 mode = 'lines+markers',
                 marker = dict(color = 'LightSeaGreen',line = dict(color = 'AliceBlue',width = 1.5))
                
                )

tr2 = go.Scatter(x = rdf2['Year'],
                 y = rdf2['Reviews'],
                 name = 'Non Fiction',
                 mode = 'lines+markers',
                 marker = dict(color = 'RoyalBlue', line = dict(color = 'AliceBlue',width = 1.5))
                
            )
data = [tr1,tr2]
layout = go.Layout(title = 'Average Reviews over the years',
                   title_x = 0.5,
                   title_font = dict(size = 16, color = 'RoyalBlue'),
                   xaxis = dict(title = 'Year'),
                   yaxis = dict(title = 'Reviews'),
                   template = 'plotly_white'
                  
                  )

fig = go.Figure(data = data, layout = layout)
fig.show()
```

* **Fiction books is higher than Non Fiction books from 2009 to 2017, but Fiction books is lower than in 2018.**
* **Fiction books is sharply descreased in 2016, and then sharply increased in 2019.**


```python
df1 = df[df['Genre']=='Fiction']
df2 = df[df['Genre']=='Non Fiction']

rdf1 = df1.groupby("Year").mean().reset_index()
rdf2 = df2.groupby("Year").mean().reset_index()

tr1 = go.Scatter(x = rdf1['Year'],
                 y = rdf1['Price'],
                 name = 'Fiction',
                 mode = 'lines+markers',
                 marker = dict(color = 'LightSeaGreen',line = dict(color = 'AliceBlue',width = 1.5))
                
                )

tr2 = go.Scatter(x = rdf2['Year'],
                 y = rdf2['Price'],
                 name = 'Non Fiction',
                 mode = 'lines+markers',
                 marker = dict(color = 'RoyalBlue', line = dict(color = 'AliceBlue',width = 1.5))
                
            )
data = [tr1,tr2]
layout = go.Layout(title = 'Average Price over the years',
                   title_x = 0.5,
                   title_font = dict(size = 16, color = 'RoyalBlue'),
                   xaxis = dict(title = 'Year'),
                   yaxis = dict(title = 'Price'),
                   template = 'plotly_white'
                  
                  )

fig = go.Figure(data = data, layout = layout)
fig.show()
```

* **Non fiction books is higher than Fiction books from 2010 to 2019.**
* **Non fiction books is sharply descreased in 2015, and then sharply increased in 2016.**

```python
df1 = df.groupby('Author').mean().sort_values('Reviews',ascending = False).reset_index().head(10)
fig = px.scatter(df1,
                 x = 'User Rating',
                 y = 'Reviews',
                 color = 'Author')

fig.update_traces(marker = dict(size = 25,
                                line = dict(color = 'rgb(0, 0, 0)',width = 1.5)),
                  selector = dict(mode = 'markers')) 

fig.update_layout(title = 'Top 10 Authors with the Highest Average Reviews',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'MidnightBlue'))
fig.show()
```

* **Top 3 Authors with the highest Average Reviews are Delia Owens,Paula Hawkins and Michelle Obama.**

```python
df1 = df.groupby(['Author','Name']).mean().sort_values('Price',ascending = False).reset_index().head(10)
fig = px.scatter(df1,
                 x = 'User Rating',
                 y = 'Price',
                 color = 'Author',
                 hover_data = ['Name']
                
                )

fig.update_traces(marker = dict(size = 20,
                                line = dict(color = 'rgb(0, 0, 0)', width = 1.5)),
                  selector = dict(mode = 'markers'))

fig.update_layout(title = 'Top 10 Authors with the Highest Average Price',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'MidnightBlue'))

fig.show()
```
