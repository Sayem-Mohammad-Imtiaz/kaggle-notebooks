### MATPLOTLIB
Matplot, verileri çizmemize yardımcı olan bir python kitaplığıdır. En kolay ve en temel grafikler çizgi, dağılım ve histogram grafikleridir.
* X ekseni zaman olduğunda çizgi grafiği daha iyidir.
* İki değişken arasında korelasyon olduğunda dağılım daha iyidir.
* Sayısal verilerin dağılımını görmemiz gerektiğinde histogram daha iyidir.
* Özelleştirme: Renkler, etiketler, çizgi kalınlığı, başlık, opaklık, ızgara, şekil boyutu, eksen çizgileri ve çizgi stili.

```python
# Line Plot
# color = color, label = label, linewidth = width of line, alpha = opacity, grid = grid, linestyle = sytle of line
data.physical_atk.plot(kind = 'line', color = 'g',label = 'physical_atk',linewidth=1,alpha = 0.8,grid = True,linestyle = ':')
#veriden hız özelliğini  
data.physical_defense.plot(color = 'r',label = 'physical_defense',linewidth=1, alpha = 0.8,grid = True,linestyle = '-.')
#veriden defens özelliğini
plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('x axis')              # label = name of label- oyuncular
plt.ylabel('y axis')              # hız ve defens özellikleri 
plt.title('Line Plot')            # title = title of plot
plt.show()
# oyuncuların fiziksel atak ve fiziksel defans derecelerini karşılaştırmış olduk
```

```python
# Scatter Plot 
# x = attack, y = defense
# verideki atak ve defans özellikleri arasındaki ilişkiyi görselleştiriyoruz
data.plot(kind='scatter', x='physical_atk', y='physical_defense',alpha = 0.5,color = 'red')
#plt.scatter(data.Attack,data.Defense,alpha = 0.5,color = 'red') bu şekilde de yazılabilir
plt.xlabel('Attack')              # label = name of label
plt.ylabel('Defence')
plt.title('Attack Defense Scatter Plot')            # title = title of plot
```

```python
# Histogram
# bins = number of bar in figure
data.mana.plot(kind = 'hist',bins = 50,figsize = (8,8))
plt.show()
```

```python
# clf() = cleans it up again you can start a fresh
data.mana.plot(kind = 'hist',bins = 50)
plt.clf()
# oluşan plotu temizliyor
# We cannot see plot due to clf()
```
