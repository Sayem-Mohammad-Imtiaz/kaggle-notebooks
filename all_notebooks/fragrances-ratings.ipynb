## Visualizations

```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mean_by_brand():
    #df['weighted'] = df.rating_score * df.votes
    #values = df.groupby('brand').weighted.sum() / df.groupby('brand').votes.sum()
    x = sorted(df.groupby('brand').rating_score.mean())
    y = range(len(x))
    
    plt.plot(x, y)
    plt.ylabel('Number of Brands')
    plt.xlabel('Rating')
    plt.title('Mean Brand Rating Distribution')

    plt.show()

mean_by_brand()
```

```python
def mean_by_gender():
    df['weighted'] = df.rating_score * df.votes
    values = df.groupby('gender').weighted.sum() / df.groupby('gender').votes.sum()
    
    objects = list(values.index)
    y_pos = range(len(objects))

    plt.bar(y_pos, values, align='center', alpha=0.5)
    plt.xticks(y_pos, objects, rotation=80)
    plt.ylabel('Rating')
    plt.title('Mean Rating by Target Demo')

    plt.show()

mean_by_gender()
```

```python
def count_by_demo():
    values = df.groupby('gender').title.nunique()
    objects = list(values.index)
    y_pos = range(len(objects))
    c = ['#3358FF','#EDAE49','#F06AA9']

    # plt.bar(y_pos, values, align='center', alpha=0.5)
    plt.pie(values, labels=objects, colors=c, autopct='%1.1f%%')
    # plt.xticks(y_pos, objects)
    # plt.ylabel('Number of products')
    plt.title('Number of products by target demographic')

    plt.show()

count_by_demo()
```

```python
def count_by_demo():
    values = df.groupby('gender').votes.sum()
    objects = list(values.index)
    y_pos = range(len(objects))
    c = ['#3358FF','#EDAE49','#F06AA9']

    # plt.bar(y_pos, values, align='center', alpha=0.5)
    plt.pie(values, labels=objects, colors=c, autopct='%1.1f%%')
    # plt.xticks(y_pos, objects)
    # plt.ylabel('Number of products')
    plt.title('Most talked-about demographic')

    plt.show()

count_by_demo()
```

```python
def count_by_brand():
    values = df.groupby('brand').title.nunique().nlargest(10)
    objects = list(values.index)
    y_pos = range(len(objects))

    plt.bar(y_pos, values, align='center', alpha=0.5)
    plt.xticks(y_pos, objects, rotation=80)
    plt.ylabel('Number of products')
    plt.title('Most prolific brands')

    plt.show()

count_by_brand()
```

```python
def votes_by_brand():
    values = df.groupby('title').votes.sum().nlargest(10)
    objects = list(values.index)
    y_pos = range(len(objects))

    plt.bar(y_pos, values, align='center', alpha=0.5)
    plt.xticks(y_pos, objects, rotation=90)
    plt.ylabel('Number of votes')
    plt.title('Most talked-about products')

    plt.show()

votes_by_brand()
```

```python
def votes_by_brand():
    masc = df.loc[df.gender=='man']
    values = masc.groupby('title').votes.sum().nlargest(10)
    objects = list(values.index)
    y_pos = range(len(objects))

    plt.bar(y_pos, values, align='center', alpha=0.5)
    plt.xticks(y_pos, objects, rotation=90)
    plt.ylabel('Number of votes')
    plt.title('Most talked-about masc products')

    plt.show()

votes_by_brand()
```

```python
def rating_curve():
    x = sorted(df.rating_score)
    y = range(len(x))
    
    plt.plot(x, y)
    plt.ylabel('Number of Products')
    plt.xlabel('Rating')
    plt.title('Product Rating Distribution')

    plt.show()

rating_curve()
```
