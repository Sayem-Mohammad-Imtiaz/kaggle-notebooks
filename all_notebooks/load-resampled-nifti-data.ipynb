#  Import and view NIfTI with SimpleITK and matplotlib

```python
import SimpleITK as sitk
import pandas as pd
import numpy as np
from pathlib import Path
from matplotlib import pyplot as plt
```

```python
DATA_DIR = Path('../input/rsna-miccai-all-resampled-to-axial-view/resampled')
df = pd.read_csv(DATA_DIR/'train_labels.csv')
fn = str(DATA_DIR/'train'/df.T2w[0]) # sitk does not support pathlib.Path
fn = fn.replace('nii.gz', 'nii') # kaggle removes the .gz from the nifti file when deflating. File should end with nii.gz instead of nii.
```

```python
img = sitk.ReadImage(fn)
nda = sitk.GetArrayFromImage(img)
nda.shape
```

```python
nda = nda.reshape(8, 8, 256, 256).transpose(0, 2, 1, 3).reshape(8*256, 8*256) # make 8 x 8 grid view
```

```python
plt.figure(figsize = (10,10))
plt.imshow(nda, interpolation='nearest', cmap='bone')
plt.show()
```
