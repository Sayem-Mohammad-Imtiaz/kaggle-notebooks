# 1. Hypothesis testing for population

I made a guess that the mean flight time of astronauts is 1440 hours (2 months, 60 days). So, the Null and Alternate hypothesis is like below. 

### H_a : 1440 = mu (claim)
### H_0 : 1440 != mu

Since we don't know the population standard deviation, we use Student's t-distribution to calculate standarized test statistic.

```python
t = (stat['mean'] - 1440) / (stat['std'] / np.sqrt(40))

print(t)
```

We calculate the critical values and decide to reject or fail to reject the null hypothesis. Critical value is calculated by scipy.

```python
# d.f = 39, a = 0.05 (level of significance)
t_c = stats.t.ppf(0.05 / 2, 39)
print(t_c) # -2.023

# Two tail test
if abs(t) > abs(t_c):
    print('reject H_0')
else:
    print('fail to reject H_0')
```

# 2. Lowering level of significance to 0.01

At #1, we decided to reject the null hypothesis at 0.05 level of significance. What will happen when we lower it to 0.01?

```python
# d.f = 39, a = 0.01
t_c = stats.t.ppf(0.01 / 2, 39)
print(t_c) # -2.708

# Two tail test
if abs(t) > abs(t_c):
    print('reject H_0')
else:
    print('fail to reject H_0')
```
