```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
data = pd.read_csv("/kaggle/input/uk-sarscov2-variant-cases-by-week/Variants.csv",parse_dates=[0],dayfirst=1,index_col=[0]).fillna(0)
```

```python
cumulative = data.cumsum()
cumulative.plot(figsize=(18,12),title="Cumulative incidence")
```

```python
incidence = data.drop("Initial count")
#print (data2)
incidence.plot(kind="line",figsize=(18,12),title="Weekly incidence")
```
