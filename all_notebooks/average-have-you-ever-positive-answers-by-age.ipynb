```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv("../input/have-you-ever/responses.csv")
columns = df.columns.tolist()
columns.remove("Age group")
```

```python
def count_trues(row):
    return sum(1 for column in columns if row[column])

df["Count"] = df.apply(count_trues, axis=1)
average = df["Count"].mean()
df = df[["Age group", "Count"]].groupby("Age group").mean()

fig = px.bar(
    x=df.index, y=df["Count"], title="Average number of interesting occurrences by age",
    labels={"x": "Age group", "y": "Average 'Yes' answer count"}, hover_name=df.index
)
fig.update_traces(hovertemplate="<b>%{x}</b><br><br>Average 'Yes' answers: %{y}")
fig.add_shape(
    type="line", line_color="salmon", line_width=3, opacity=1, line_dash="dot",
    x0=0, x1=1, xref="paper", y0=average, y1=average, yref="y"
)
fig.show()
```
