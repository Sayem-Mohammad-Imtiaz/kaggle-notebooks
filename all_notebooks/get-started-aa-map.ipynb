```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

ds = pd.read_csv("../input/fishing-data-north-atlantic/full_dataset.csv")
raster = pd.read_csv("../input/fishing-data-north-atlantic/raster.csv")
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
co = world[world.name.isin(["United Kingdom","Iceland","Norway","Ireland"])].set_geometry('geometry')
raster = gpd.GeoDataFrame(raster)
```

```python
test = ds[ds.month==ds.sample().month.values[0]][["patch","surf_temp"]]
test.index = test.patch
dict_ = test.to_dict()["surf_temp"]
raster["val"] = raster.LOKREF.replace(dict_)
```

```python
from shapely import wkt

raster['geometry'] = raster['geometry'].apply(wkt.loads)
raster.set_geometry('geometry');
```

```python
base = co.plot(color='teal', edgecolor='black', figsize=(16, 9))

raster.plot(ax=base,column='val', cmap='coolwarm', edgecolor='k', legend=False)
```

```python
co = co.to_crs("epsg:3574")
raster = raster.set_crs("epsg:4326")
raster = raster.to_crs("epsg:3574")
```

```python
base = co.plot(color='teal', edgecolor='black', figsize=(16, 9))

raster.plot(ax=base,column='val', cmap='coolwarm', edgecolor='k', legend=False)
```
