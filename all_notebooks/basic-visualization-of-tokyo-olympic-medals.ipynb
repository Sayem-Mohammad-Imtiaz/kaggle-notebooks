# Reading the dataset

```python
df = pd.read_csv('../../kaggle/input/2021-olympics-medals-in-tokyo/Tokyo Medals 2021.csv')
df.head()
```

# Importing the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.offline as pyo
import plotly.graph_objs as go
from plotly.offline import iplot
import seaborn as sns
%matplotlib inline
```

# Medals by Country 

```python
top=df.sort_values(by="Rank By Total",ascending=True)[:11]
top
```

# All countries by Medal stats

```python
figure = plt.figure(figsize=(30,20))
sns.barplot(y="Country", x="Total",data=df)
plt.xticks()
plt.xlabel('Total Medal')
plt.ylabel('Country')
plt.title('Country by Total Medal')
plt.show;
```

# Top 10 Country by Total medals

```python
plt.figure(figsize=(30,10))
data = [go.Bar(x=top['Country'],y=top['Total'])]
layout=go.Layout(title='Medals')
fig=go.Figure(data=data,layout=layout)
#pyo.plot(fig)
iplot(fig)
```

# The 10 Country Rewards by Gold Medals

```python
figure = plt.figure(figsize=(10,6))
sns.barplot(y="Country", x="Gold Medal",data=top)
plt.xticks()
plt.xlabel('Gold Medal')
plt.ylabel('Country')
plt.title('Top 10 Country by Gold Medal')
plt.show;
```

# The 10 Country Rewards by Silver Medals

```python
figure = plt.figure(figsize=(10,6))
sns.barplot(y="Country", x="Silver Medal",data=top)
plt.xticks()
plt.xlabel('Silver Medal')
plt.ylabel('Country')
plt.title('Top 10 Country by Silver Medal')
plt.show;
```

# The 10 Country Rewards by Bronze Medals

```python
figure = plt.figure(figsize=(10,6))
sns.barplot(y="Country", x="Bronze Medal",data=top)
plt.xticks()
plt.xlabel('Bronze Medal')
plt.ylabel('Country')
plt.title('Top 10 Country by Bronze Medal')
plt.show;
```

```python
values=top['Country']
names=top['Gold Medal']
figure = plt.figure(figsize=(10,10))
plt.pie(names, labels=values, labeldistance=1.15);
plt.show();
```

# Comparing Top 10 Countries by their medal!!

```python
#NESTED!!
trace1 = go.Bar(
    x=top['Country'],  
    y=top['Gold Medal'],
    name = 'Gold',
    marker=dict(color='#FFD700') 
)
trace2 = go.Bar(
    x=top['Country'],
    y=top['Silver Medal'],
    name='Silver',
    marker=dict(color='#9EA0A1') 
)
trace3 = go.Bar(
    x=top['Country'],
    y=top['Bronze Medal'],
    name='Bronze',
    marker=dict(color='#CD7F32') 
)
data = [trace1, trace2, trace3]
layout = go.Layout(
    title='2021 Olympic Medals by Country'
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
