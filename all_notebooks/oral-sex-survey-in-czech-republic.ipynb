The first time I saw this data, I was very curious about it and I decided to analyze it and see what information I can get.
First of all, I wanted to see the distribution of age being asked in this survey and level of education that they have.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#read the csv file.
df = pd.read_csv('D:\WORK\Mis_programas\Oral_sex_survey\oral_sex_eng_translation.csv', header=0)

#amount of people
plot1 = plt.figure(1)
x = df['Age']
plt.grid(axis='x', alpha=0.75)
plt.xlabel('Age')
plt.ylabel('Counts')
plt.title('Amount of people')
plt.xticks(np.arange(10, 70, 5))

hist_plot = plt.hist(x, bins = 36, ec='black')

for i in range(36):
    plt.text(hist_plot[1][i], hist_plot[0][i], str(hist_plot[0][i]))
    
plot1 = plt.figure(4)
x = df['Education']
hist_plot = plt.hist(x, bins=np.arange(5)-0.5, ec='black')
plt.ylabel('Counts')
plt.xticks(size = 8, rotation=0)

plot1 = plt.figure(5)
df_men = df.loc[df['Sex'] == 'Male']
x = df_men['Do you engage in oral sex?']

df_women = df.loc[df['Sex'] == 'Female']
y = df_women['Do you engage in oral sex?']

plt.hist(x, alpha=0.7, bins=np.arange(7)-0.7, rwidth=0.4, color='black', label='men', ec='black')
plt.hist(y, alpha=0.7, bins=np.arange(7)-0.3, rwidth=0.4, color='red', label='women', ec='black')

plt.legend(loc='upper right')
plt.title('Do you engage in oral sex?')
plt.ylabel('Counts')
plt.xticks(size = 8, rotation=90)

#COMBINED HISTOGRAM 2
plot1 = plt.figure(6)
df_men = df.loc[df['Sex'] == 'Male']
x = df_men['Kind of sex and eye contact']

df_women = df.loc[df['Sex'] == 'Female']
y = df_women['Kind of sex and eye contact']

plt.hist(x, alpha=0.7, bins=np.arange(8)-0.7, rwidth=0.4, color='black', label='men', ec='black')
plt.hist(y, alpha=0.7, bins=np.arange(8)-0.3, rwidth=0.4, color='red', label='women', ec='black')

plt.legend(loc='upper right')
plt.title('Kind of sex and eye contact')
plt.ylabel('Counts')
plt.xticks(size = 8, rotation=90)
```
