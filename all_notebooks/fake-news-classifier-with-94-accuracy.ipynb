# LABELLING OF DATA

```python
news.label=pd.get_dummies(news['label'])['FAKE']
```

```python
y=news['label']
```

```python
X=news.drop('label',axis=1)
```

# LEMMATIZING THE DATASET 

```python
import nltk
```

```python
from nltk.corpus import stopwords
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
import re
```

```python
corpus=[]
for i in range(len(X)):
    rev=re.sub('[^a-zA-Z]',' ',X['text'][i])
    rev=rev.lower()
    rev=rev.split()
    rev=[lemmatizer.lemmatize(word) for word in rev if word not in stopwords.words('english')]
    rev=' '.join(rev)
    corpus.append(rev)
```

```python
len(corpus)
```

# VECTORIZATION OF DATA

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_v=TfidfVectorizer(max_features=8000,ngram_range=(1,3))
X=tfidf_v.fit_transform(corpus).toarray()
```

```python
# tfidf_v.get_feature_names()
```

```python
X=pd.DataFrame(X,columns=tfidf_v.get_feature_names())
```

```python
X
```

# TRAIN TEST SPLIT OF DATASET

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

# PREDICTION USING NAIVE_BAYES MODEL

```python
from sklearn.naive_bayes import MultinomialNB
classi=MultinomialNB()
```

```python
classi.fit(X_train,y_train)
```

```python
y_pred=classi.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.heatmap(cm,annot=True,fmt='.4g',xticklabels=['Real','Fake'],yticklabels=['Real','Fake'])
```

```python
classification_report(y_test,y_pred)
```

# PREDICTION USING PASSIVEAGGRESSIVECLASSIFIER

```python
from sklearn.linear_model import PassiveAggressiveClassifier
pac=PassiveAggressiveClassifier()
pac.fit(X_train,y_train)
```

```python
y_pac=pac.predict(X_test)
```

```python
accuracy_score(y_test,y_pac)
```

```python
cm=confusion_matrix(y_test,y_pac)
```

```python
sns.heatmap(cm,annot=True,fmt='.4g',xticklabels=['Real','Fake'],yticklabels=['Real','Fake'])
```

```python
classification_report(y_test,y_pac)
```
