# 1. Read csv into dataframe

```python
csv_file = '/kaggle/input/iemocap-emotion-speech-database/iemocap_full_dataset.csv'

iemocap_df = pd.read_csv(csv_file)
iemocap_df.head()
```

# 2. Sample EDA

## 2.1 Emotion Distribution, Scripted and Improvised sessions

```python
impro = iemocap_df.loc[iemocap_df.method == 'impro']
script = iemocap_df.loc[iemocap_df.method == 'script']

plt.figure(figsize=(12,6))

plt.subplot(1,2,1)
script.emotion.value_counts().plot.pie(autopct='%1.1f%%',label='',pctdistance=0.85)
plt.title('Scripted')

plt.subplot(1,2,2)
impro.emotion.value_counts().plot.pie(autopct='%1.1f%%',label='',pctdistance=0.85)
plt.title('Improvised')
```

## 2.2 Four emotions in improvised session, commonly used in speech emotion recognition (SER) evaluation.

```python
impro = iemocap_df.loc[iemocap_df.method == 'impro']
emo = ['neu','hap','sad','ang']
impro_selected = impro.loc[impro.emotion.isin(emo)]
impro_selected.emotion.value_counts().plot.pie(autopct='%1.1f%%',label='',pctdistance=0.85)
```
