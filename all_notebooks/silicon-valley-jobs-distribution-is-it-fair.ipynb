```python
from aif360.datasets import StandardDataset
from aif360.metrics import BinaryLabelDatasetMetric
dataset_orig = StandardDataset(
    df=job,
    categorical_features=['demographics'],
    label_name=['job_category'],
    features_to_keep=['percentage','demographics','job_category'],
    protected_attribute_names=['demographics'],
    favorable_classes=['Executives'], # for output values that are favorable historically
    privileged_classes=privileged_group # for protected attributes which have advantage historically
)

dataset_orig_train, dataset_orig_test = dataset_orig.split([0.7], shuffle=True)
```
