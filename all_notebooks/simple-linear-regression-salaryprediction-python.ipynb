```python
#Read data and view first few rows
data = pd.read_csv("/kaggle/input/salary-data-simple-linear-regression/Salary_Data.csv")
data.head()
```

```python
# describe data

data.describe()
```

```python
# get predictor & target variable
x = data[["YearsExperience"]]
y = data[["Salary"]]
```

```python
# Split data into treain and test

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
# Predict 
from sklearn.linear_model import LinearRegression

model =LinearRegression()

model.fit(X_train,y_train)

y_predict = model.predict(X_test)


from sklearn.metrics import r2_score
print("R-square metric : ", r2_score(y_test,y_predict))
```
