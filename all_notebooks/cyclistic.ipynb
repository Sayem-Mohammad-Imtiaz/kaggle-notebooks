```python
my_mode <- function(x) {                     
  unique_x <- unique(x)
  tabulate_x <- tabulate(match(x, unique_x))
  unique_x[tabulate_x == max(tabulate_x)]
    }
```

```python
mode_weekday <- my_mode(total_bike_clean$day_of_week)
```
