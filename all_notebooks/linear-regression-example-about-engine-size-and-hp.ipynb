# ****Hello in this notebook we will make linear regression beetween engine size and horsepower.

# First we start by importing our dataset

```python
df = pd.read_csv('../input/car-price-prediction/CarPrice_Assignment.csv')
df.head()
```

# Then we assing value Engine size to X. HorsePower to Y


```python
hp = df.horsepower
engSize= df.enginesize


x = engSize
y = hp
x= x.values.reshape(-1,1)





```

# We split our data by 1/3 for test 2/3 to train our linear regressino model.

```python
x_train, x_test, y_train, y_test = ms.train_test_split(x,y,test_size=1/3,random_state = 0)
```

# Here we train our model

```python
reg = lr.LinearRegression()
reg.fit(x_train,y_train)
```

# Here we predic our new y values

```python
y_pred = reg.predict(x_test)
```

# Here we callculate our linear regression score its %60. it means our data  is not suited very well to the linear regression analysis. But it can be used to predict approximately values.

```python
score = mt.r2_score(y_test,y_pred)
print(score)
```

# Here we can see how our model shows the relation of increase in enginesize and horsepower.

```python
plt.scatter(x_test,y_test, color = 'b')
plt.plot(x_test,y_pred, color = 'r')
plt.show()
```
