```python
# K NEIGHBORS CLASSIFIER ON 'heart_disease' DATASET TO COMPARE THE ORIGINAL DATA WITH THE PREDICTED DATA
# ACCURACY DECREASES WHEN RELATABLE COLUMNS ARE REMOVED 

import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn import model_selection
from sklearn.metrics import accuracy_score

df=pd.read_csv('../input/heart-disease-patients-details/heart_disease.csv')

Y=df['result']
X=df.loc[:,'age':'thal']

X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,test_size=0.33,random_state=4)

knn=KNeighborsClassifier()
knn.fit(X_train,Y_train)
abc=knn.predict(X_test)

print('The Accuracy Score is {}'.format(knn.score(X_train,Y_train)))  # ACCURACY OF TRAINED DATA
print(abc)
print(Y_test)
print(accuracy_score(Y_test,abc)) # ACCURACY OF OBTAINED DATA
```
