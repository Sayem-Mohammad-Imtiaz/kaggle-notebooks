## Do smokers pay more in medical costs?

```python
g=sns.catplot(data=df, kind='swarm', x='age', y='charges', hue='sex', col='smoker', palette='Spectral').set_xticklabels(rotation=90)
```

- Overall smokers experience higher medical costs
