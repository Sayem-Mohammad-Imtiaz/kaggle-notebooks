```python
import numpy as np
import pandas as pd
from glob import iglob
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
stats = pd.read_csv('../input/vtuber-livechat/chat_stats.csv')
sc_stats = pd.read_csv('../input/vtuber-livechat/superchat_stats.csv')
channels = pd.read_csv('../input/vtuber-livechat/channels.csv')

# merge stats columns
stats = stats.query('channelId == "UCMwGHR0BTZuLsmjY_NT5Pwg"')
sc_stats = sc_stats.query('channelId == "UCMwGHR0BTZuLsmjY_NT5Pwg"')
stats = pd.merge(stats, sc_stats, on=['channelId', 'period'], how='left')
numeric_columns = stats.select_dtypes(include=['number']).columns
stats[numeric_columns] = stats[numeric_columns].fillna(0).astype(
        'int')
stats.drop(columns=['channelId'], inplace=True)

channels = channels.query('channelId == "UCMwGHR0BTZuLsmjY_NT5Pwg"')
channels.drop(columns=['channelId', 'name.en', 'affiliation', 'group'], inplace=True)
```

```python
periods = stats['period']
px.bar(stats.query('period > "2021-01"'),
       x='period',
       y='chatCount',
       color='chatNunique',
       color_continuous_scale=px.colors.sequential.Purples,
       text='chatCount',
       labels={
           'chatCount': '<b>Chats</b>',
           'chatNunique': '<b>Unique Chatters</b>',
           'period': 'Period'
       },
       title="How much chats are posted on Ina's stream?"
      )
```

```python
px.bar(stats.query('period > "2021-01"'),
       x='period',
       y='deletionCount',
       color_continuous_scale=px.colors.sequential.Purples,
       text='deletionCount',
       labels={
           'banCount': 'Number of <b>BANs</b>',
           'deletionCount': 'Number of <b>Deleted Chats</b>',
           'chatCount': 'Number of <b>Chats</b>',
           'chatNunique': 'Number of <b>Unique Chatters</b>',
           'period': 'Period'
       },
       title="How many chats were deleted from Ina's stream?"
).update_layout(barmode='group')
```
