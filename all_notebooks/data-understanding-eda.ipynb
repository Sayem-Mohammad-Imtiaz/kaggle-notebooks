**for eda i mapped event_type as original steps to int as ordinal encoding**

```python
cleanup_nums = {"event_type":     {"seen_in_list": 1, "seen": 2, "suggest-new": 3, 
                                    'suggest_similar': 4, 'sent_catalog_link': 5, 'visit_request-new':6, 'visit_request-canceled':7,
                                    'visit-new':8, 'visit-canceled':9, 'visit-unsuccess':10, 'visit-success':11,
                                    'meeting_request-new':12, 'meeting_request-canceled':13, 'meeting-new':14,'meeting-canceled':15, 
                                   'meeting-unsuccess':16, 'meeting-success':17, 'deal-success':18}}
                                                          
log.replace(cleanup_nums, inplace=True)
```

**Unique numbers of user_id and item_id**

```python
log.user_id.unique().size, log.item_id.unique().size
```

```python
p5s = (log[(log['event_type'] ==18) ][['item_id', 'user_id']])
#u5s = list(log[log.event_type==18]['user_id'])


pu5s = log[(log.item_id.isin(p5s.item_id.values)) & (log.user_id.isin(p5s.user_id.values))]

##pu5s.shape
```

### **Fastest Rents among 54 success deals**

```python
plt.style.use('seaborn-talk')
diff = []
for item in list(pu5s.item_id.unique()):
    
    d = pd.to_datetime(pu5s[pu5s.item_id==item].sort_values(by='create_timestamp')['create_timestamp']).min()
    f = pd.to_datetime(pu5s[pu5s.item_id==item].sort_values(by='create_timestamp')['create_timestamp']).max()
    diff.append(tuple((item, f-d)))
    
df = pd.DataFrame(diff, columns =['item', 'time']) 
  
df 
```

```python
df.time = df['time'] / np.timedelta64(1, 's')

df.sort_values(by='time', inplace=True)
```

```python
plt.figure(figsize=(10, 15))
plt.title('Rented by Second in the System Log')
plt.style.use('seaborn-talk')
sns.barplot(y=df.item, x=df.time)
```

### **Success deals properties and steps**

```python
plt.style.use('seaborn-talk')
steps = []
for item in list(pu5s.item_id.unique()):
    
    d = pu5s[pu5s.item_id==item].sort_values(by='create_timestamp')['event_type'].values
    
    steps.append(tuple((item, d)))
    
df = pd.DataFrame(steps, columns =['item', 'step']) 
  
df
```
