```python
gold_price = pd.read_csv('/kaggle/input/gold-and-silver-prices-dataset/gold_price.csv')
silver_price = pd.read_csv('/kaggle/input/gold-and-silver-prices-dataset/silver_price.csv')
```

```python
gold_price = gold_price.set_index('date')
silver_price = silver_price.set_index('date')
metals_price = pd.merge(gold_price, silver_price, how='inner', 
                        left_index=True, right_index=True, suffixes=['_gold', '_silver'])
```

```python
metals_price.head()
```

```python
metals_price.corr()
```

```python
_ = metals_price[['price_gold', 'price_silver']].plot(secondary_y='price_silver', 
                                                      color=['gold', 'silver'], 
                                                      title='prices per ounce in USD',
                                                      figsize=(20, 10))
```
