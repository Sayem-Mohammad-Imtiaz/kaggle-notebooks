solver = pywraplp.Solver('SolverAtividadeSAD', pywraplp.Solver.GLOP_LINEAR_PROGRAMMING)

carvao = solver.NumVar(0, solver.infinity(), 'carvao')
picanha = solver.NumVar(0, solver.infinity(), 'picanha')
wisky = solver.NumVar(0, solver.infinity(), 'wisky')
problema = solver.NumVar(0, solver.infinity(), 'problema')

obj1 = solver.Constraint(10, 30)
obj1.SetCoefficient(carvao, 1)
obj1.SetCoefficient(picanha, 2)

obj2 = solver.Constraint(10, 30)
obj2.SetCoefficient(carvao, 1)
obj2.SetCoefficient(problema, 2)

obj3 = solver.Constraint(10, 50)
obj3.SetCoefficient(wisky, 1)
obj3.SetCoefficient(problema, 2)

objective = solver.Objective()
objective.SetCoefficient(carvao, 1)
objective.SetCoefficient(picanha, 4)
objective.SetCoefficient(wisky, 4)
objective.SetCoefficient(problema, 2)
objective.SetMaximization()

solver.Solve()
print('Número de variáveis =', solver.NumVariables())
print('Número de restrições =', solver.NumConstraints())

print('Solução:')
print('carvao =', carvao.solution_value())
print('picanha =', picanha.solution_value())
print('wisky =', wisky.solution_value())
print('problema =', problema.solution_value())
