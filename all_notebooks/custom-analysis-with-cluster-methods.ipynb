# Kmeans with elbow and silhouette score

```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import seaborn as sns
from sklearn.metrics import silhouette_score
```

in low dimensions, only consider the last two coloums fisrt
further analysis: label the gender and use PCA to reduce dim

```python
# load data and see examples
dataset = pd.read_csv('../input/customer-segmentation-tutorial-in-python/Mall_Customers.csv')
X1=dataset.iloc[:,3:]
X2=dataset.iloc[:,3:]
dataset.head(5)
```

```python
X1.iloc[:,0].plot(kind = 'hist')
plt.xlabel('Annual Income')
plt.show()
```

```python
X1.iloc[:,1].plot(kind = 'hist')
plt.xlabel('Spending score')
plt.show()
```

## 1.elbow method to find the optimal clusternumber

```python
#Elbow method

cost=[]
for i in range (1,10):
    kmeans=KMeans(n_clusters=i,init='k-means++', random_state = 10)
    kmeans.fit(X1)
    cost.append(kmeans.inertia_)
    
plt.plot(range(1,10),cost)
plt.title('The elbow method')
plt.xlabel('The number of Clusters')
plt.ylabel('cost function')
plt.show()
```

```python
#Here the optimal value of k is 5.
kmeans=KMeans(n_clusters=5,init='k-means++', random_state = 10)
cluster_labels= kmeans.fit_predict(X1)
print(cluster_labels)
X1['predict1'] = cluster_labels
```

```python
# Clusters predicted with k-Means
sns.FacetGrid(X1,hue='predict1').map(plt.scatter, "Annual Income (k$)", "Spending Score (1-100)").add_legend();
```

## 2. Silhouette score analysis

```python
from sklearn.metrics import silhouette_score

range_n_clusters = [2, 3, 4, 5, 6, 7, 8]

for num_clusters in range_n_clusters:
    
    # intialise kmeans
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50,random_state= 10)
    kmeans.fit(X2)
    
    cluster_labels = kmeans.labels_
    
    # silhouette score
    silhouette_avg = silhouette_score(X2, cluster_labels)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
#K-means with k=5 clusters

cluster = KMeans(n_clusters=5, max_iter=150, random_state= 20)
cluster.fit(X2)
```

```python
# Cluster labels

cluster.labels_
```

```python
X2['predict2']=cluster.labels_
X2.head()
```

```python
X2.predict2.value_counts(ascending = True)
```
