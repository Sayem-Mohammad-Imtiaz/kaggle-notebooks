```python
eruption=pd.read_csv('/kaggle/input/volcanic-eruptions/database.csv')
eruption
```

```python
median_longitude=np.median(eruption['Longitude'])
print("Медианное значение долготы: ",median_longitude)
```

```python
import statistics as stat

mode_longitude=stat.mode(eruption['Longitude'])
print("Мода долготы: ",mode_longitude)
```

```python
mean_longitude=np.mean(eruption['Longitude'])
print("Среднее значение долготы: ",mean_longitude)
```

```python
min_longitude=np.min(eruption['Longitude'])
print("Минимальное значение долготы: ",min_longitude)
```

```python
max_longitude=np.max(eruption['Longitude'])
print("Максимальное значение долготы: ",max_longitude)
```

```python
new_eruption=eruption.copy()
new_eruption.set_index('Number', inplace=True)

new_eruption["diff"]=new_eruption.max(axis=1)-new_eruption.min(axis=1)
new_eruption
```

```python
eruption['Region'].value_counts()
```

```python
South_and_central_America=eruption[(eruption['Region']=='South America') | 
                                   (eruption['Region']=='México and Central America') | 
                                   (eruption['Region']=='Hawaii and Pacific Ocean') | 
                                     (eruption['Region']=='West Indies')]
count_South_and_central_America=len(South_and_central_America)

Africa_and_Oceans=eruption[(eruption['Region']=='Africa and Red Sea') | 
                           (eruption['Region']=='Mediterranean and Western Asia') | 
                           (eruption['Region']=='Atlantic Ocean') | 
                     (eruption['Region']=='Middle East and Indian Ocean')]
count_Africa_and_Oceans=len(Africa_and_Oceans)

Asia=eruption[(eruption['Region']=='Japan, Taiwan, Marianas') | 
        (eruption['Region']=='Kamchatka and Mainland Asia') | 
              (eruption['Region']=='Kuril Islands')]
count_Asia=len(Asia)

Oceania_and_Antarctica_and_South_Asia=eruption[(eruption['Region']=='Indonesia') | 
                                               (eruption['Region']=='Melanesia and Australia') | 
                                               (eruption['Region']=='Philippines and SE Asia') | 
                                               (eruption['Region']=='New Zealand to Fiji') | 
                                               (eruption['Region']=='Antarctica')]
count_Oceania_and_Antarctica_and_South_Asia=len(Oceania_and_Antarctica_and_South_Asia)

North=eruption[(eruption['Region']=='Canada and Western USA') | 
               (eruption['Region']=='Alaska') | 
               (eruption['Region']=='Iceland and Arctic Ocean')]
count_North=len(North)


import matplotlib.pyplot as plt
values=[count_South_and_central_America, count_Africa_and_Oceans, count_Asia, count_Oceania_and_Antarctica_and_South_Asia, count_North]
labels=['Южная и Центральная Америка, Гавайи, Вест-Индия', 'Африка, Средиземноморье, Средний Восток, Атлантический океан', 
        'Япония, Тайвань, Марианские острова, Камчатка, Курильские острова', 'Страны Океании, Южная Азия, Антарктида', 
        'Канада, Ирландия, Аляска, Северный Ледовитый Океан']

fig, ax = plt.subplots()
ax.pie(values, labels=labels, autopct='%1.2f%%')
ax.axis("equal")
```

```python
fig=plt.figure(figsize=(10,6))
plt.hist(South_and_central_America['Activity Evidence'], color='green')
plt.title("Доказательство деятельности")
ax.set_xlabel('Вид доказательства')
ax.set_ylabel('Количество извержений')
```
