```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**Importing Data from Dataset**

```python
df = pd.read_csv("../input/spotify-dataset-19212020-160k-tracks/data.csv")
df.head(3)
```

**Listing the columns present in Dataset**

```python
col = df.columns
print(col)
```

**Finiding out the artists in the dataset**

```python
art = df['artists']
print(art)
```

**Top 10 Most Popular Artists**

```python
plt.figure(figsize=(10, 3))
ax = df.groupby("artists")["popularity"].sum().sort_values(ascending=False).head(10).plot(kind='bar')
ax.set_title("TOP 10 artists")
ax.set_ylabel("POPULARITY")
ax.set_xlabel("ARTIST")
plt.show()
```

**Top 10 Most Unpopular Artist**

```python
plt.figure(figsize=(10, 3))
ax = df.groupby("artists")["danceability"].sum().sort_values(ascending=False).head(10).plot(kind='bar')
ax.set_title(" 10 Unpopular  artists")
ax.set_ylabel("DANCEABILITY and ENERGY")
ax.set_xlabel("ARTIST")
plt.show()
```

**Chnages in Audio characteristics over years**

```python
plt.figure(figsize=(32, 16))
sns.set(style="whitegrid")
columns = ["acousticness","danceability","energy","speechiness","liveness","valence"]
for col in columns:
    x = df.groupby("year")[col].mean()
    ax= sns.lineplot(x=x.index,y=x,label=col)
ax.set_title('Audio characteristics over year')
ax.set_ylabel('Measure')
ax.set_xlabel('Year')
```
