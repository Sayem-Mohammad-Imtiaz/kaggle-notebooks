```python
# from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
# import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

Data = pd.read_csv('../input/netflix-shows/netflix_titles.csv')
Features = ['type', 'title', 'director', 'cast']
for feature in Features:
    Data[feature] = Data[feature].fillna('')


def combine_features(row):
    return row['title'] + row['cast'] +row['description']


Data["combined_features"] = Data.apply(combine_features, axis=1)
cv = CountVectorizer()
count_matrix = cv.fit_transform(Data["combined_features"])
cosine_sim = cosine_similarity(count_matrix)


def find_title_from_index(index):
    return Data[Data.show_id == index]["title"].values[0]


def find_index_from_title(title):
    return Data[Data.title == title]["show_id"].values[0]


print('Please Enter your desired movie or TV show')
movie_name = input()

while not ((Data['title'] == movie_name).any()):
    print('Please search again and use more general keywords')
    movie_name = input()

movie_index = find_index_from_title(movie_name)

movie_index = int(movie_index[1:])
similar_movies = list(enumerate(cosine_sim[movie_index]))
sorted_similar_movies = sorted(similar_movies, key=lambda x: x[1], reverse=True)[1:]
i = 0
print(movie_name)
for element in sorted_similar_movies:
    ID = 's' + str(element[0])
    print(find_title_from_index(ID))
    i = i + 1
    if i > 5:
        break
```
