```python
# 1 SVC (Support Vector Classifier)
import numpy as np
import pandas as pd 

#import required libraries
import matplotlib.pyplot as plt     

#dataset connection
dataset= pd.read_csv('../input/healthcaredataset/healthcare-dataset-stroke-data.csv')      

#feature data
x=dataset.iloc[:,[4,5]].values     
#target data
y=dataset.iloc[:,6].values         

from sklearn.model_selection import train_test_split     #train data & test data separation
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.25,random_state=0)

#data scaling
from sklearn.preprocessing import StandardScaler         
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.fit_transform(x_test)


from sklearn.svm import SVC
classifier = SVC()
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)

#confusion matrix for comparison of prediction
from sklearn.metrics import confusion_matrix             
cm = confusion_matrix(y_test,y_pred)

#to test accuracy
from sklearn.metrics import confusion_matrix,accuracy_score   
cm = confusion_matrix(y_test,y_pred)
accuracy_score(y_test,y_pred)

print(cm)
print(accuracy_score(y_test,y_pred))

#to see graphical view
from matplotlib.colors import ListedColormap                  
x_set, y_set = x_test, y_test
X1, X2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)

plt.title('SVC')
plt.xlabel('Glucose Lavel & BMI')
plt.ylabel('Strok')
plt.legend()
plt.show()
```

```python
# 3 DecisionTreeClassifier model

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset= pd.read_csv('../input/healthcaredataset/healthcare-dataset-stroke-data.csv')

#----partitioning feature and target--------

x=dataset.iloc[:, [4,5]].values
y=dataset.iloc[:,6].values

#------split train and test---------
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=.25,random_state=0)


#------making standard format-------

from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.fit_transform(x_test)
x_train


from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier()

classifier.fit(x_train,y_train)

#-----prediction----------
y_predict= classifier.predict(x_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,y_predict)
print(cm)
print(accuracy_score(y_test,y_predict))
#-----drawing
plt.title('Decision tree model (test set)')
plt.xlabel('Glucose Lavel Vs BMI')
plt.ylabel('Strok')
#------copy code---
from matplotlib.colors import ListedColormap
x_set, y_set = x_test, y_test
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)


plt.title('Decision Tree (Test set)')
plt.xlabel('Glucose Lavel Vs BMI')
plt.ylabel('Strok')
plt.legend()
plt.show()
```

```python
# 4 RandomForestClassifier model

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset= pd.read_csv('../input/healthcaredataset/healthcare-dataset-stroke-data.csv')

#----partitioning feature and target--------

x=dataset.iloc[:, [4,5]].values
y=dataset.iloc[:,6].values

#------split train and test---------
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=.25,random_state=0)


#------making standard format-------

from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x_train=sc_x.fit_transform(x_train)
x_test=sc_x.fit_transform(x_test)
x_train


from sklearn.ensemble import RandomForestClassifier
classifier=RandomForestClassifier(n_estimators=100)

classifier.fit(x_train,y_train)

#-----prediction----------
y_predict= classifier.predict(x_test)

from sklearn.metrics import confusion_matrix, accuracy_score
cm=confusion_matrix(y_test,y_predict)
print(cm)
print(accuracy_score(y_test,y_predict))
#-----drawing
plt.title('RandomForestClassifier model (test set)')
plt.xlabel('Glucose Label Vs BMI')
plt.ylabel('Strok')
#------copy code---
from matplotlib.colors import ListedColormap
x_set, y_set = x_test, y_test
x1, x2 = np.meshgrid(np.arange(start = x_set[:, 0].min() - 1, stop = x_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = x_set[:, 1].min() - 1, stop = x_set[:, 1].max() + 1, step = 0.01))
plt.contourf(x1, x2, classifier.predict(np.array([x1.ravel(), x2.ravel()]).T).reshape(x1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(x1.min(), x1.max())
plt.ylim(x2.min(), x2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(x_set[y_set == j, 0], x_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)


plt.title('RandomForestClassifier (Test set)')
plt.xlabel('Glucose Lavel Vs BMI')
plt.ylabel('Strok')
plt.legend()
plt.show()
```
