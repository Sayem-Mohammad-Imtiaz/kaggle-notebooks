## Importing the libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

## Read/ Explore the data

```python
df = pd.read_csv('../input/european-football-database-20192020/D1.csv')
```

```python
f"Data has {df.shape[0]} Rows and {df.shape[1]} Columns"
```

```python
df.isna().sum().sum()
```

```python
df.head()
```

## Time column in home and away maches

```python
# Get the time for Bayern Munich Maches
y = df['Time'][df['HomeTeam'] == 'Bayern Munich'].value_counts()
x = df['Time'][df['AwayTeam'] == 'Bayern Munich'].value_counts()
# Make them a data frame
df_ = pd.DataFrame([y,x]).T
# Rename the columns
df_.columns=['Home Time', 'Away Time']
df_
```

## How many goals Bayern Munich scored (100)

```python
# Get the away and home goals
x = df['FTHG'][df['HomeTeam'] == 'Bayern Munich'].value_counts()
y = df['FTAG'][df['AwayTeam'] == 'Bayern Munich'].value_counts()
# Make them as a data frame
df_ = pd.DataFrame([x,y]).T
# Rename the columns
df_.columns = ['Home Goals', 'Away Goals']
# Fill the NaN values with 0
df_.fillna(0)
```

```python
plt.figure(figsize=(10,6))
plt.plot(df_['Home Goals'], marker='o', linestyle=':', label='Home Goals', color='purple')
plt.plot(df_['Away Goals'], marker='o', linestyle=':', label='Away Goals', color='orange')
plt.title('Bayern Munich 100 Goals', size=20)
plt.ylabel('How many times', size=15)
plt.xlabel('Goals', size=15)
plt.legend()
plt.show()
```

## Bayern Munich Result

```python
fig, ax = plt.subplots(1,2,figsize=(15,6))
plt.suptitle('H=Home Win, D=Draw, A=Away Win', fontsize=15)
ax[0].set_title('Home Results', size=20)
ax[1].set_title('Away Results', size=20)
sns.countplot(x=df['FTR'][df['HomeTeam'] == 'Bayern Munich'], ax=ax[0])
sns.countplot(x=df['FTR'][df['AwayTeam'] == 'Bayern Munich'], ax=ax[1])
plt.tight_layout(2.0)
plt.show()
```
