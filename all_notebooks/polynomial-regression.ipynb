**Polynomial Regression (To predict salaries based on previous job position)** -
To check if the person is speaking the truth or bluffing.

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing dataset
ds = pd.read_csv('../input/positionsalaries-data/position-salaries data.csv')
x = ds.iloc[:,1:-1].values
y = ds.iloc[:,-1].values
```

Position and Level column, both show the position of the individual's previous job, so to avoid categorical data in position column, we use only Level column as the independent variable.

```python
x = ds.iloc[:, 1:-1].values
y = ds.iloc[:, -1].values
print(x)
print(y)
```

We do not split the data since in the dataset there are only 10 values. So, splitting the data further into test and train would decrease the accuracy.

```python
#training the linear regression model on the dataset
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x,y)
```

```python
#training the polynomial regression model on the dataset
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree = 5)
x_poly = pf.fit_transform(x)
```

```python
#applying linear regression on the polynomial set
from sklearn.preprocessing import PolynomialFeatures
lr = LinearRegression()
lr.fit(x_poly,y)
```

```python
#visualising polynomial regression results
plt.scatter(x, y, color = 'red')
plt.plot(x, lr.predict(x_poly))
plt.title('Truth or Bluff')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

```python
#checking accuracy
from sklearn.metrics import r2_score
r2_score(y, lr.predict(x_poly))
```

**Accuracy = 99.9%**
