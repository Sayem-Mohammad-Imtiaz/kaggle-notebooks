```python
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/zillow-1/house_data.csv')

df['Size'] = df['Size'].str.replace('sqft', '').str.replace('-','').str.replace('lot','').str.strip()
df['Bathrooms'] = df['Bathrooms'].str.replace('ba','').str.replace('-','').str.strip()
df['Bedrooms'] = df['Bedrooms'].str.replace('bds', '').str.replace('-','').str.strip()
df.fillna(0, inplace=True)
df.dropna(inplace=True)

home_df = df[((df['Sale Status'].str.lower().str.contains('home', na=False)) | (df['Sale Status'].str.lower().str.contains('house', na=False)))]

home_df.fillna('0', inplace=True)
```

```python
#Narrow down houses by one state.
texas_df = home_df[home_df['Address'].str.lower().str.contains('tx')]
```

```python
#Function that takes in state abbr. as a parameter and returns avg. house price.
def narrow_state(state : str):
    df1 = home_df[home_df['Address'].str.lower().str.contains(f'{state}')]
    print(f'Average house price in {state.upper()} : ${df1["Raw Price"].mean()}')
    #df1.to_csv(f'./state_houses/{state}_homes.csv', index=False)
```
