```python
df = pd.read_csv('/kaggle/input/notcias-publicadas-no-brasil/Historico_de_materias.csv')
clear_output()
df.assunto.value_counts().plot.pie()
```

```python
df.dropna(inplace=True)
df.info()
```

```python
df.titulo[1]
```

```python
df.assunto.value_counts()
```

```python
data = df[['titulo','assunto']].copy()
data.assunto  =  data.assunto.astype('category')
data.rename(columns={'titulo': 'noticia'},inplace=True)
assunto_cats = data.assunto.cat.categories
assunto_cats
```

```python
print(data.shape)
data.info()
```

```python
data.assunto   = data.assunto.cat.codes
```

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(data,test_size=0.2)
print('train.shape',train.shape)
print('test.shape',test.shape)
```

```python
from autoviml.Auto_NLP import Auto_NLP
train_nlp, test_nlp, transform_nlp,preds = Auto_NLP('noticia',
                                              train,
                                              test,'assunto',
                                              score_type='balanced_accuracy',
                                              modeltype='Classification',
                                              top_num_features=100,
                                              verbose=2,
                                              build_model=True)
```

```python
print('AutoNLP Pipeline:')
transform_nlp 
```

```python
predics = transform_nlp.predict(test['noticia'])
```

```python
acc_metric = tf.keras.metrics.Accuracy()
acc_metric.update_state(test['assunto'].values,predics)
acc = acc_metric.result().numpy()
print(f'Accuracy:\t{acc}')
```
