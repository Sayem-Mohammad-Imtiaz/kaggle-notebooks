# Demonstration for PCA with Python

```python
import numpy as np, pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
iris= pd.read_csv("/kaggle/input/iris-flower-dataset/IRIS.csv")
iris.head()
```

```python
y = iris.pop("species")
iris.head()
```

```python
iris.describe()
```

# Scaling the data

```python
from sklearn.preprocessing import PowerTransformer
scaler = PowerTransformer()
```

```python
X = iris.copy()
X.head()
```

```python
x = scaler.fit_transform(X)
x
```

```python
x[:5,:5]
```

# Applying PCA on the data

```python
from sklearn.decomposition import PCA
pca = PCA(random_state=42)
```

```python
pca.fit(x)
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
plt.bar(range(1,len(pca.explained_variance_ratio_)+1), pca.explained_variance_ratio_)
plt.show()
```

```python
var_cumu = np.cumsum(pca.explained_variance_ratio_)
var_cumu
```

# Making the Scree plot

```python
plt.figure(figsize=(10,5))
plt.plot(range(1,len(var_cumu)+1), var_cumu)
plt.show()
```

## PCA with 2 components

```python
pc2 = PCA(n_components=2, random_state=42)
newiris = pc2.fit_transform(x)
newiris
```

```python
newiris.shape
```

### Making a dataframe out of it for convenience



```python
df = pd.DataFrame(newiris, columns=["PC1", "PC2"])
df.head()
```

```python
df_final = pd.concat([df, y], axis=1)
df_final.head()
```

## Plotting the data

```python
import seaborn as sns
sns.pairplot(data=df_final, x_vars=["PC1"], y_vars=["PC2"], hue = "species", size=8)
plt.show()
```
