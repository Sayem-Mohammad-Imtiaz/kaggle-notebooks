```python
import numpy as np
import pandas as pd
```

```python
#Read Dataset
df = pd.read_csv('../input/hotel-booking/hotel_booking.csv')
```

```python
#Check the data
df.head()
```

```python
#Check information
df.info()
```

```python
#Check the number of rows
pd.DataFrame([df.shape],columns= ['Rows','Columns'],index= ['Check the number of rows'])
```

```python
#Is there data missing? If yes, which column has the highest number? Has it lost data?
number_null = df.isnull().sum().sum()
print('There are {} missing data in the dataset'.format(number_null))
max_null = df.isnull().sum().idxmax()
print('The most missing data is {}'.format(max_null))    
```

```python
#Remove the 'company' column from the dataframe.
df = df.drop('company', axis=1)
df.head()
```

```python
#Which countries do most travelers come from? Identify the top 5 countries. 
df["country"].value_counts().head()
```

```python
#Name the person who has the highest paid Rate Daily Average (ADR) How much is?
_ADR = df[['name','adr']].groupby('name').sum().sort_values('adr', ascending=False) [0:1]
print(_ADR)
```

```python
#Specify the average of total ADRs. Up to decimal places
print("The Average Number Of ADR Is : {}".format(round(df["adr"].mean(),2)))
```

```python
#Specify the average number of nights spent. Up to two decimal places
SumNights = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
print("The average number of nights is : {}.".format(round(SumNights.mean(),2)))
```

```python
#Specify the name and email address of people who had 5 special requests.
SR = df[['name','email','total_of_special_requests']]
SR[SR['total_of_special_requests']>4]
```

```python
#What are the most common surnames? Identify 5 large families do
df['name'].apply(lambda lname:lname.split()[1]).value_counts().head().to_frame('value_counts')
```

```python
#Identify the people who have booked the hotel with the largest number of babies and children do
print("{} >>> number of children :".format(df.name[df['children'].idxmax()]),("{}.".format(df['children'].max())))
print("{} >>> number of babies :".format(df.name[df['babies'].idxmax()]),("{}.".format(df['babies'].max())))
```
