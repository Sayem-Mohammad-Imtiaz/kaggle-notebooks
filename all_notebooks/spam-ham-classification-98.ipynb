```python
df = pd.read_csv('/kaggle/input/spam-or-not-spam-dataset/spam_or_not_spam.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df['label'].value_counts()
```

```python
spam = df[df['label']== 1]
spam.head()
```

```python
ham = df[df['label']== 0]
ham.head()
```

```python
ham.shape, spam.shape
```

```python
ham = ham.sample(spam.shape[0])
```

```python
ham.shape, spam.shape
```

```python
data = ham.append(spam, ignore_index=True)
data.tail()
```

```python
X = data['email'].fillna(' ')
y = data['label']
import re
def preprocess(text):
    text = text.str.lower()
    text = text.str.replace(r'\r', ' ')
    text = text.str.replace(r'\n', ' ')
    text = text.str.replace('[^a-zA-Z0-9]', ' ')
    return text
```

```python
x = preprocess(X)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.3, random_state=0, shuffle = True, stratify= data['label'])
xtrain
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vec = TfidfVectorizer()
xtrain = vec.fit_transform(xtrain)
```

```python
xtrain.shape
```

```python
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
clf = Pipeline([('tfidf', TfidfVectorizer()),('clf', RandomForestClassifier(n_estimators=100, n_jobs=-1))])
clf.fit(xtrain, ytrain)
```

```python
ypred = clf.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
confusion_matrix(ytest, ypred)
```

```python
classification_report(ytest, ypred)
```

```python
accuracy_score(ytest, ypred)
```
