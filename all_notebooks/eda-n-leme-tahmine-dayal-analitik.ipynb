GaussianNB modelini yaratalım ve eğitelim.

```python
nb = GaussianNB()
nb_model = nb.fit(X_train, y_train)
```

****K-Nearest Neighbors (KNN) Modelini Eğitme****


K-Nearest Neighbors (KNN) modelini yaratalım ve eğitelim.

```python
knn_params = {"n_neighbors": np.arange(2,50)}
```

2 ile 50 arasında olan her sayıyı komşu sayısı olarak deneyecek ve en iyi komşu değerini göreceğiz.

```python
knn = KNeighborsClassifier()
knn_cv = GridSearchCV(knn, knn_params, cv = 3)
knn_cv.fit(X_train, y_train)
```

```python
print("En iyi skor: " + str(knn_cv.best_score_))
print("En iyi parametreler: " + str(knn_cv.best_params_))
```

En iyi komşu değerinin 13 olduğu gözlenmektedir. 

```python
knn = KNeighborsClassifier(13)
knn_tuned = knn.fit(X_train, y_train)
```

Veri çerçevesindeki tüm X_test gözlemlerini tahmin ettirip y_pred değişkenine atayalım.

```python
y_pred = knn_tuned.predict(X_test)
```

Doğruluk skorunu görüntüleyelim.

```python
accuracy_score(y_test, y_pred)
```

* Yüzde %92 oranında doğruluk oranı vardır.

Karmaşıklık matrisini görüntüleyelim.

```python
KarmaşıklıkMatrisi4 = confusion_matrix(y_test,y_pred)
KarmaşıklıkMatrisi4
```

Komşuluk değerlerine göre doğruluk skorunun değişimini görselleştirelim.

```python
score_list = [] 

for each in range(2,50,1): 
    knn2 = KNeighborsClassifier(n_neighbors = each)
    knn2.fit(X_train,y_train)
    score_list.append(knn2.score(X_test, y_test))

plt.plot(range(2,50,1),score_list)
plt.xlabel("K Değerine En Yakın Komşu Sayıları")
plt.ylabel("Doğruluk Skoru")
plt.show()
```

* 2 den 50`ye kadar olan komşu sayılarını for döngüsünde dönüp 49 adet k değeri için doğruluk skorlarını görselleştirmiş olduk.
