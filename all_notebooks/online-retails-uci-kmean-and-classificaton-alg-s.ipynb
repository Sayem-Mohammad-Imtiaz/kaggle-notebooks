```python
null_df=df[df['Description'].isnull()]
```

```python
sto_cod=null_df['StockCode']
```

```python
stockcodes=sto_cod.values
```

```python
indexes=sto_cod.index
```

```python
codes=list(stockcodes)
```

```python
codes=set(codes)
```

```python
codes=list(codes)
```

```python
len(codes)
```

```python
desc={}
stock_with_null_desc=[]
indexes_with_null_desc=[]
for i in range(len(codes)):
    founded_df=df[df['StockCode'].str.contains(codes[i])]
    founded_df=founded_df.dropna() 
    if (len(founded_df['Description'].mode())==0):
        indexes_with_null_desc.extend(df[df['StockCode']==codes[i]].index)
        stock_with_null_desc.append(codes[i])
        continue
    desc[codes[i]]=founded_df['Description'].mode()[0]
    


    
    
    
```
