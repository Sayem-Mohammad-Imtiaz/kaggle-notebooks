```python
#bedakan variabel x dan y
x = df.drop(columns='Y_harga_unit')
y = df['Y_harga_unit']
```

```python
#split data menjadi training dan testing dengan porso 80:20
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)
```

```python
#membuat variabel linear regresi
lin_reg = LinearRegression()
```

```python
#training data
lin_reg.fit(x_train, y_train)
```

```python
#cetak nilai koefisien dan intersep
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
coef_dict = {
    'fitur':x.columns,
    'koefisien': lin_reg.coef_
}
coef = pd.DataFrame(coef_dict, columns=['fitur', 'koefisien'])
coef
```

```python
y_pred = lin_reg.predict(x_test)
```

```python
lin_reg.score(x_test, y_test)
```
