# Data Processing
to complete the lost data, drop the irrelevent data, and creat some new data

```python
print ("Before", train.shape, test.shape)
train = train.drop(['Ticket', 'Cabin', 'Name', 'PassengerId'], axis = 1)
test = test.drop(['Ticket', 'Cabin', 'Name', 'PassengerId'], axis = 1)
print("After",train.shape, test.shape )
```

While "Ticket", "Name", "PassengerI"' and "Cabin" lost too much data or too complex to find the relation, we choose to drop these data

```python
train['Sex'] = train['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
test['Sex'] = test['Sex'].map( {'female': 1, 'male': 0} ).astype(int)
train.head()
```

Converting "Sex" to numerics 0 and 1

```python
train.Age.value_counts().sort_values(ascending = False).head(5)
```

```python
test.Age.value_counts().sort_values(ascending = False).head(5)
```

```python
train.Age = train.Age.fillna(24)
test.Age = test.Age.fillna(24)
```

```python
test.Age.unique()
```

We replace the lost data of "Age" by the most possible age: 24 both in train dataset and test dataset.

```python
train.head()
```

```python
train['Embarked'] = train['Embarked'].map( {'S': 1, 'C': 2, 'Q':3, np.nan: 0, 24: 0} ).astype(int)
test['Embarked'] = test['Embarked'].map( {'S': 1, 'C': 2, 'Q':3, np.nan: 0, 24: 0} ).astype(int)
train.head()
```

"Embarked" converting to numerics

```python
train['FamilySize'] = pd.DataFrame(train.SibSp + train.Parch)
train.describe()
```

```python
test['FamilySize'] = pd.DataFrame(test.SibSp + test.Parch)
test.describe()
```

```python
train= train.drop(['SibSp', 'Parch'], axis = 1)
test= test.drop(['SibSp', 'Parch'], axis = 1)
train.head()
```

```python
train.Age.unique()
```

```python
test.Fare.value_counts().head()
```

```python
test.Fare = test.Fare.fillna(7.75)
```

Sum the "SibSp" and "Parch" as "FamilySize"

# Modeling traning
* Logistic Regression
* KNN or k-Nearest Neighbors
* Support Vector Machines
* Naive Bayes classifier
* Decision Tree
* Random Forrest
* Perceptron
* Artificial neural network
* Relevance Vector Machine

```python
x_train = train.drop("Survived", axis = 1)
y_train = train.Survived
x_test = test
x_train.shape,y_train.shape, x_test.shape
```

```python
x_test.info()
```

* Logistic Regression

```python
logreg = LogisticRegression()
logreg.fit(x_train, y_train)
y_pred = logreg.predict(x_test)
acc_log = round(logreg.score(x_train, y_train) * 100, 2)
acc_log
```

* k-NN

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)
acc_knn = round(knn.score(x_train, y_train) * 100, 2)
acc_knn
```

* Support Vector Machines

```python
svc = SVC()
svc.fit(x_train, y_train)
y_pred = svc.predict(x_test)
acc_svc = round(svc.score(x_train, y_train) * 100, 2)
acc_svc
```

* Naive Bayes classifier

```python
gaussian = GaussianNB()
gaussian.fit(x_train, y_train)
y_pred = gaussian.predict(x_test)
acc_gaussian = round(gaussian.score(x_train, y_train) * 100, 2)
acc_gaussian
```

* Perceptron

```python
perceptron = Perceptron()
perceptron.fit(x_train, y_train)
y_pred = perceptron.predict(x_test)
acc_perceptron = round(perceptron.score(x_train, y_train) * 100, 2)
acc_perceptron
```

* Linear SVC

```python
linear_svc = LinearSVC()
linear_svc.fit(x_train, y_train)
y_pred = linear_svc.predict(x_test)
acc_linear_svc = round(linear_svc.score(x_train, y_train) * 100, 2)
acc_linear_svc
```

* Decision Tree

```python
decision_tree = DecisionTreeClassifier()
decision_tree.fit(x_train, y_train)
y_pred = decision_tree.predict(x_test)
acc_decision_tree = round(decision_tree.score(x_train, y_train) * 100, 2)
acc_decision_tree
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(x_train, y_train)
y_pred = random_forest.predict(x_test)
random_forest.score(x_train, y_train)
acc_random_forest = round(random_forest.score(x_train, y_train) * 100, 2)
acc_random_forest
y_pred.astype(np.int32)
np.savetxt("output.csv", y_pred)
```

* Stochastic Gradient Descent

```python
sgd = SGDClassifier()
sgd.fit(x_train, y_train)
y_pred = sgd.predict(x_test)
acc_sgd = round(sgd.score(x_train, y_train) * 100, 2)
acc_sgd
```

# A brife comparation of these models

```python
models = pd.DataFrame({
    'Model': ['Support Vector Machines', 'KNN', 'Logistic Regression', 
              'Random Forest', 'Naive Bayes', 'Perceptron', 
              'Stochastic Gradient Decent', 'Linear SVC', 
              'Decision Tree'],
    'Score': [acc_svc, acc_knn, acc_log, 
              acc_random_forest, acc_gaussian, acc_perceptron, 
              acc_sgd, acc_linear_svc, acc_decision_tree]})
models.sort_values(by='Score', ascending=False)
```
