```python
#reading data set, checking column names, data type, and null values, checking a big overview of data.
data = pd.read_csv("../input/google-job-skills/job_skills.csv")
data.info()
data.describe()
```

#Questions we are trying to answer
1. Which jobs are most popular at Google?
2. Which category is most popular at Google?
3. Which location advertises the most number of jobs?
4. Which minimum qualificatios are most in-demand at Google?
5. Which preferred qualification are most in-demand at Google?


#### Question 1 - Which jobs are most popular at Google?

```python

pp = data['Title'].value_counts()
jobs = pd.DataFrame(pp)
import matplotlib.pyplot as plt
jobs.head(10).plot.bar(legend='', figsize=(12,8),title='Most advertised jobs')
plt.show()
```

#### Question 2. Which category advertises the most number of jobs at Google?


```python
aa = data['Category'].value_counts()
category = pd.DataFrame(aa)
category.head(10).plot.bar(legend='', figsize=(12,8), title='Most advertised categories')
plt.show()
```

#### Question 3. Which location advertises most jobs?


```python
bb = data['Location'].value_counts()
location = pd.DataFrame(bb)
location.head(10).plot.bar(legend='', figsize=(12,8), title='Locations with the most number of advertised jobs')
plt.show()
```

#### Question 4.Which minimum qualificatios are most in-demand at Google?


```python
cc = data['Minimum Qualifications'].value_counts()
min_qua = pd.DataFrame(cc)
min_qua.head(10)
```

#### Question 5. Which preferred qualification are most in-demand at Google?


```python
dd = data['Preferred Qualifications'].value_counts()
pref_qua = pd.DataFrame(dd)
pref_qua.head(10)
```
