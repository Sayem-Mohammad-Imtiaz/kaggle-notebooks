## Intro

Since there are quite a few categorical variables and the result itsself that is DEATH_EVENT is categorical I am running a Chi Square Tests to find relationships between each column and death events if any.

Following are the columns I am using for tests

* age
* anaemia
* diabetes
* high_blood_pressure
* sex
* smoking


```python


import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as ws
from scipy.stats import chi2_contingency
```

```python
df = pd.read_csv("../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv")
df.head()
```

```python
chi2_contingency(pd.crosstab(index=df['age'], columns=df['DEATH_EVENT']))
```

```python
chi2_contingency(pd.crosstab(index=df['anaemia'], columns=df['DEATH_EVENT']))
```

```python
chi2_contingency(pd.crosstab(index=df['diabetes'], columns=df['DEATH_EVENT']))
```

```python
chi2_contingency(pd.crosstab(index=df['high_blood_pressure'], columns=df['DEATH_EVENT']))
```

```python
chi2_contingency(pd.crosstab(index=df['sex'], columns=df['DEATH_EVENT']))
```

```python
chi2_contingency(pd.crosstab(index=df['smoking'], columns=df['DEATH_EVENT']))
```

## Inference:

The only column categorical that has relationship with Death Events is ***age***, that we can also see in graph belo

```python
plt.figure(figsize=(20,6))
title = plt.title('Survival and Deaths by Age', fontsize=20)
title.set_position([0.5, 1.15])
ax = sns.countplot(x="age", hue="DEATH_EVENT", data=df)
ax.set_xlabel('Age')
ax.set_ylabel('Count')
a = ax.set_xticklabels(ax.get_xticklabels(), rotation=0, horizontalalignment='center')
```
