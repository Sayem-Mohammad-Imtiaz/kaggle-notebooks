```python
city = []
      
for i in lista["city"]:
        if i not in city:
            city.append(i)
del(city[2])
print(city)
```

```python
houses = []
copiaCity = []
for i in lista["city"]:
    copiaCity.append(i)
houses.append(copiaCity.count(city[0]))
houses.append(copiaCity.count(city[1]))
houses.append(copiaCity.count(city[2]))
houses.append(copiaCity.count(city[3]))
houses.append(copiaCity.count(city[4]))
print(houses)
```

```python
plt.pie(houses, labels=city, autopct='%1.2f%%')
plt.title('Porcentagem de Casas para Alugar em cada Cidade')
plt.show()
```

```python
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("../input/aula-2-ia-dataset/CasasParaAlugar.csv", index_col=0)
# Selecionei um subconjunto das variáveis de entrada para fins de simplificação
FEATURES_ANALISAR = ['city', 'area', 'rooms', 'bathroom', 'parking spaces',  'floor', 'animal', 'furniture',]
df = df.loc[:, FEATURES_ANALISAR]
df
```

```python
print("{} linhas\n{} colunas/features.".format(df.shape[0], df.shape[1]))
```

```python
# Funcao do Pandas usada para contar o numero de valores vazios de cada coluna
data = df.isna().sum(axis=0)
y = list(range(df.shape[1]))
x = data.values

# Criamos uma figura
fig, ax = plt.subplots(figsize=(15, 10))

# Plota as barras
ax.barh(y=y, width=x)

# Adiciona informações no gráfico
ax.set_yticks(y)
ax.set_yticklabels(df.columns.values)
ax.set_title("Quantidade de variáveis ausentes por coluna")
plt.show()
```
