```python
x = dataset.drop(['charges'],axis =1)
y = dataset.charges.values
```

## Encoding Categorical Data

Changing the similer string values into numberical values for higher accuracy

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct = ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[1,4,5])],remainder='passthrough')
x = np.array(ct.fit_transform(x)) # train test model accept array data so we have converted our data into array form using numpy
```

## Training Model on training set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

# Random Forest Regression


```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(x_train, y_train)
```

## Model Predictions

```python
y_pred = regressor.predict(x_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

## Accuracy

```python
print('Accuracy {:.2f}%'.format(regressor.score(x_test,y_test)*100))
```

```python
print("RMSE valuse for # Random Forest Regression:", sqrt(mse(y_test, y_pred)))
```
