# Weight for each Features
                                Ranges     Weights
* Happiness Scores              1 - 10      9
* Social Support                0.1 - 1     8
* Healthy Life Expectancy       0 - 100     9
* Freedom to Make Life Choices  0.1 - 1     9
* Generocity                    -0.9 - 0.9  8
* Perceptions of Corruption     0.1 - 0.9   3

```python
weights = [9,8,8,9,7,6]
```

```python
studpd['Social Support'] = studpd['Social Support'].apply(lambda x : x * 10)
```

```python
studpd['Healthy Life Expectancy'] = studpd['Healthy Life Expectancy'].apply(lambda x : x / 10)
```

```python
studpd['Freedom to Make Life Choices'] = studpd['Freedom to Make Life Choices'].apply(lambda x : x * 10)
```

```python
studpd['Generosity'] = studpd['Generosity'].apply(lambda x : x * 10)
```

```python
studpd['Perceptions of Corruption'] = studpd['Perceptions of Corruption'].apply(lambda x : x * 10)
```

```python
studpd['to_live'] = (studpd['Happiness Scores'] * weights[0] + studpd['Social Support'] * weights[1] + studpd['Healthy Life Expectancy'] * weights[2] + studpd['Freedom to Make Life Choices'] * weights[3] + studpd['Generosity'] * weights[4] - studpd['Perceptions of Corruption'] * weights[5])/ sum(weights)
```

```python
studpd.sort_values(by=['to_live'],ascending=False)
```
