```python
pokemon = pd.read_csv('/kaggle/input/pokemon/pokemon.csv')
pokemon.columns
```

```python
pokemon = pokemon[['name', 'hp', 'speed']]
pokemon.head()
```

```python
pokemon.hist()
```

```python
import plotly.graph_objects as go
```

```python
fig = go.Figure(layout=go.Layout(
        title=go.layout.Title(text="Health and Speed of Pokemon"),
        xaxis_title="Health",
        yaxis_title="Speed",
    ))

fig.add_trace(go.Scatter(x=pokemon.hp, y=pokemon.speed, mode='markers', hovertext=pokemon.name))
fig.show()
```

```python
hp_normalized = (pokemon.hp - pokemon.hp.mean()) / pokemon.hp.std()
speed_normalized = (pokemon.speed - pokemon.speed.mean()) / pokemon.speed.std()
pokemon['hp_normalized'] = hp_normalized
pokemon['speed_normalized'] = speed_normalized
pokemon.head()
```

```python
pokemon[['hp_normalized', 'speed_normalized']].hist()
```

```python
pokemon['outlier'] = pokemon.apply(lambda row: int((row.hp_normalized < -1.5) or 
                                                   (row.hp_normalized > 2.5) or 
                                                   (row.speed_normalized < -2) or 
                                                   (row.speed_normalized > 2.5)), axis=1)
pokemon.head()
```

```python
fig = go.Figure(layout=go.Layout(
        title=go.layout.Title(text="Health and Speed of Pokemon"),
        xaxis_title="Health",
        yaxis_title="Speed",
    ))

outlier_pokemon = pokemon[pokemon.outlier == 1]
normal_pokemon = pokemon[pokemon.outlier == 0]
fig.add_trace(go.Scatter(x=outlier_pokemon.hp, y=outlier_pokemon.speed, mode='markers', hovertext=outlier_pokemon.name, name='Outliers'))
fig.add_trace(go.Scatter(x=normal_pokemon.hp, y=normal_pokemon.speed, mode='markers', hovertext=normal_pokemon.name, name='Normal'))
fig.show()
```
