### Conversion of all elements to lables

```python
k = df.keys()
for i in k:
    ob = LabelEncoder()
    n = str(i) + '_n'
    df[n] = ob.fit_transform(df[i])
df
```

#### Where:
##### Form V1_n to V9_n --->  Assign b = 0, o=1 and x = 2
##### And v10_n         ---> negetive = 0 , positive = 1 

### Assignin dependent and independent values to variables

```python
X = df.iloc[:,10:-1]
Y = df['V10_n']
X
```

### Splitting data into train and test data

```python
xtrain,xtest,ytrain,ytest = train_test_split(X,Y,test_size=0.3)
```

### Training model with data

```python
model = dc()
model.fit(X,Y)
```
