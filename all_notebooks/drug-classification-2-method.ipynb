```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split as trn
from sklearn.preprocessing import LabelEncoder
import os
```

```python
df = pd.read_csv("../input/drug-classification/drug200.csv")
print (df.columns.tolist())
```

```python
df.head(10)
```

**Merubah Data Menjadi Numerik**

```python
dataframe_int = df.copy()

# Membuat objek yang bernama encoder
encoder = LabelEncoder()
categorical_data = ['Sex','BP','Cholesterol','Drug']
for kolom in categorical_data:
    dataframe_int[kolom] = encoder.fit_transform(df[kolom])
```

```python
dataframe_int.head()
```

# **Memisahkan Data**

Karena pada kasus ini kita ingin memprediksi data yang ada pada kolom Drug, maka kita pisahkan antara data tersebut menjadi 2 data

```python
data = dataframe_int.drop('Drug',axis=1)
label = dataframe_int['Drug']
```

```python
data.head()
```

# **Melakukan Split Model**

memisahkan data menjadi 2 data yaitu data latihan dan data tes dengan modul Sklearn

```python
x_train, x_test, y_train, y_test = trn(data,label,test_size=15,random_state=42,stratify=label)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

# **SVM**

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.svm import LinearSVC #classifier SVM
from sklearn.metrics import classification_report, confusion_matrix
```

```python
model = SVC(gamma='scale')
model.fit(x_train,y_train)
prediction = model.predict(x_test)
```

```python
prediction
```

```python
model.score(x_test,y_test)
```

# **KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_X_train = scaler.fit_transform(x_train)
scaled_X_test = scaler.transform(x_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=1, p=2 )
```

```python
knn.fit(scaled_X_train,y_train)
```

```python
pred_k = knn.predict(scaled_X_test)
pred_k
```

```python
knn.predict_proba(scaled_X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,pred_k))
```

```python
print(classification_report(y_test,pred_k))
```

```python
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(y_test, pred_k)
accuracy
```
