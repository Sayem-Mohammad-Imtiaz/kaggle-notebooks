```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
df = pd.read_csv("../input/pima-indians-diabetes-database/diabetes.csv")
```

```python
df.head()
```

```python
def sign(x):
    res = 0 if x <= 0 else 1
    return res
```

```python
def hypothesis(x,w):
    res = 0
    for i in range(len(x)):
      res += x[i]*w[i]

    res = sign(res)
    return res
```

```python
x_data = df.iloc[:,:-1].values # first 8 features
x_data = np.append(x_data,np.zeros([len(x_data),1]),1) # adding bias
y_data = df.iloc[:,-1].values # target values
l_rate = 0.1 #learning rate
```

```python
x_data.shape
```

```python
for i in range(len(x_data)):
    x_data[i][8] = 1
```

```python
import random
weights=[]
for i in range(0,9):
    n = random.randint(0,10)
    weights.append(n/10)
print(weights)
```

```python
for k in range(0,100): #100 epochs
    flag = True
    print("Epoch:",k)
    for i in range(len(x_data)):
        y_predict = hypothesis(x_data[i],weights)
        print("x = ", x_data[i], "\ty_predict = ",y_predict,"\ty_actual",y_data[i])
        if y_predict != y_data[i]:
            flag = False
            for j in range(len(weights)):
                w_temp = weights[j] + l_rate*(y_data[i] - y_predict)*x_data[i][j]
                weights[j] = w_temp
    if flag==True: #break if all the predicted values are right because fully optimized
        break
    print("\n\n")
```

```python
print("weights = ",weights)
weights = np.array(weights)
weights.shape
```

```python
#Test
count=0
for i in range(len(x_data)):
    y_predict = hypothesis(x_data[i],weights)
    if y_predict == y_data[i]:
        count+=1
print("Accuracy:",(count*100)/len(x_data))
    
```
