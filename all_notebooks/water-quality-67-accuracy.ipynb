## Finding Null Values

As it turns out there are many NaN values in ph, Sulfate, and Trihalomenthanes column 

```python
sns.heatmap(dataset.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

___
## Finding Outliers

Before I fill NaN values, I need to know the outliers of those 3 columns, so that I can decide whether to fill the NaN values with the mean's or the median's value of the column

Here I'm using 2 types of figures:
1. Boxplot (You can easily notice the outliers from the circles)
2. Histogram

```python
fig = plt.figure(figsize =(5, 3))
# Creating plot
plt.boxplot(dataset["ph"].dropna())
# show plot
plt.show()
```

```python
sns.distplot(dataset["ph"].dropna(),kde=False,color='darkred',bins=40)
```

```python
fig = plt.figure(figsize =(5, 3))
# Creating plot
plt.boxplot(dataset["Sulfate"].dropna())
# show plot
plt.show()
```

```python
sns.distplot(dataset["Sulfate"].dropna(),kde=False,color='darkred',bins=40)
```

```python
fig = plt.figure(figsize =(5, 3))
# Creating plot
plt.boxplot(dataset["Trihalomethanes"].dropna())
# show plot
plt.show()
```

```python
sns.distplot(dataset["Trihalomethanes"].dropna(),kde=False,color='darkred',bins=40)
```

___
## Filling the NaN values

Here you can use both Mean or Median to fill the NaN values, both ended up resulting the same accuracy.

```python
dataset["ph"] = dataset["ph"].fillna((dataset["ph"].mean()))
dataset["Sulfate"] = dataset["Sulfate"].fillna((dataset["Sulfate"].mean()))
dataset["Trihalomethanes"] = dataset["Trihalomethanes"].fillna((dataset["Trihalomethanes"].mean()))

sns.heatmap(dataset.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

___
## Deleting the outliers

```python
Q1 = dataset.quantile(0.25)
Q3 = dataset.quantile(0.75)
IQR = Q3 - Q1
dataset = dataset[~((dataset < (Q1 - 1.5 * IQR)) |(dataset > (Q3 + 1.5 * IQR))).any(axis=1)]
```

___
## Split the Data

```python
x = dataset.iloc[:,0:9].values
y=dataset.iloc[:,9:].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

### Don't forget to scale the x values

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

# Logistic Regression

Resulted in 0.63 accuracy

```python
from sklearn.linear_model import LogisticRegression
log=LogisticRegression()
log.fit(x_train,y_train)
```

```python
log_pred=log.predict(x_test)
```

```python
log_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(y_test,log_pred)
print("Logistic Regression accuracy value: {:.2f}".format(accuracy))
```

# Decision Tree Classifier

Resulted in 0.60125 accuracy


```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion="entropy", random_state=0)
dt.fit(x_train,y_train)
```

```python
dt_predict=dt.predict(x_test)
```

```python
dt_predict
```

```python
accuracy2=accuracy_score(y_test,dt_predict)
print("Decision Tree Classifier accuracy value: {:.2f}".format(accuracy2))
```

# KNN (K-Nearest Neighbor)

Resulted in 0.655 accuracy

```python
from sklearn.neighbors import KNeighborsClassifier
kkn = KNeighborsClassifier(n_neighbors=19)
kkn.fit(x_train,y_train)
```

```python
kkn_pred=kkn.predict(x_test)
```

```python
accuracy3=accuracy_score(y_test,kkn_pred)
print("KNN accuracy value: {:.2f}".format(accuracy3))
```

# Naive Bayes

Resulted in 0.62 accuracy

```python
from sklearn.naive_bayes import GaussianNB
naive = GaussianNB()
naive.fit(x_train,y_train)
```

```python
naive_pred=naive.predict(x_test)
```

```python
accuracy4=accuracy_score(y_test,naive_pred)
print("Naive Bayes accuracy value: {:.2f}".format(accuracy4))
```

# Random Forest Classifier

Resulted in 0.67 accuracy

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
rf_pred=rf.predict(x_test)
```

```python
accuracy5=accuracy_score(y_test,rf_pred)
print("Random Forest Classifier accuracy value: {:.2f}".format(accuracy5))
```
