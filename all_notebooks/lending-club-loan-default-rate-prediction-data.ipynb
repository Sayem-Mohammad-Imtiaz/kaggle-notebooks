```python
df = df[df.loan_status != 'Current']
df = df[df.loan_status != 'In Grace Period']
df = df[df.loan_status != 'Late (16-30 days)']
df = df[df.loan_status != 'Late (31-120 days)']
df = df[df.loan_status != 'Does not meet the credit policy. Status:Fully Paid']
df = df[df.loan_status != 'Does not meet the credit policy. Status:Charged Off']
df = df[df.loan_status != 'Issued']
## The Loan status is the Target Variable which I am going to use for the model building the main goal is to predict the Loan default
## and Non default status so in the column we have unneccessary categories which are not usefull for the model building thats 
## why we are dropping those categories
```

```python
df['loan_status'] = df['loan_status'].replace({'Charged Off':'Default'})
## As we are going to predict the Default and Non-Default from the dataset , we have a category called charged off which means 
##"Charge off" means that the credit grantor wrote your account off of their receivables as a loss, and it is closed to future 
##charges. When an account displays a status of "charge off," it means the account is closed to future use, although the debt 
##is still owed.
## That means when bank doent receives monthly payment to clear they debt they will delecar that the account holder is not paying
## the owned by the bank for past 3 or 4 months (not exactly to be according to the bank norms) they delcare that this account 
## will be charged off and comes under to the Default Category thats why we are merging them.
```

```python
df['verification_status']= df['verification_status'].replace({'Source Verified':'Verified'})
## As we can see that in the dataset the verification status column has three category which are source verified , verified and 
## Not verified. We can observe that verified and source verified are belongs to same category .Thats why I merged them into 
## one category has verified.
```

```python
df['issue_d']=pd.to_datetime(df['issue_d']).dt.year
## The issue date is a cloumns that represents the year that the loan was granted to the appliant 
##This column is in object data type has the date columns has to be datetime format. Thats why I am 
##Changing the issue_date column datatype object to integer by changing into year
```

```python
df["emp_length"] = df["emp_length"].replace({'years':'','year':'',' ':'','<':'','\+':'','n/a':'0'}, regex = True)
df["emp_length"] = df["emp_length"].apply(lambda x:int(x))
## we can see in the dataset that emp_length is in odject datatype has it to be in the numerical format tp the 
## that object to numerical type there are some symbols with those we cant change the datatype so we are removing the symbols 
## from the data and then chaning them into integer datatype
```

```python
df['home_ownership'].replace(['NONE', 'ANY','OTHER'], 'RENT', inplace=True)
## we can see in the home ownership column we have multiple categories in that we have None , Any , Other which are similar to
## Rent because the person who doent have own house or apartment they have to live in rent tahts why we merged those columns
```

```python
df['grade'] = df['grade'].map({'A':7,'B':6,'C':5,'D':4,'E':3,'F':2,'G':1})

## to reduce columns so its not ordinal data thats why i given a number to the grade category 
```

```python
df.reset_index(inplace=True)

#after dropping the huge junk of rows the index of the dataset has changed thats why I am resting the index for dataset 
```

### Categorical Input and Categorical Output ---- Statistical Test

#### H0 : The Feature is not Significant to target variable
#### H1 : The Feature is Significant to target variable

```python
import scipy.stats as st
for i in cat_cols.columns:
    ct=pd.crosstab(df['loan_status'],df[i])
    print('chi-square test for ',i,'p-value is ', st.chi2_contingency(ct)[1])
    print('\n')
    
    ## we are doing chi square test because we are comparing the Categorical Independent and Categorical dependent to check
    ## the statistical significance of the variables in the dataset 
    ## The p-value is less that 0.05 so we are rejecting the H0 Null Hypothesis
```

### Numerical Input and Categorical Output ---- Statistical Test
