Linear Regression to predict House Prices

Splitting the data into Train & Test

```python
x=df.drop(columns=['Lon','Lat','Address','Zip','Price'])
y=df['Price']
```

```python
print(x.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
from math import sqrt
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
```

```python
print("Accuracy of training dataset:",model.score(x_train,y_train))
print("Accuracy of test dataset:",model.score(x_test,y_test))
```

```python
print('Coefficient of determination: %.2f' % r2_score(y_test, predict))
```
