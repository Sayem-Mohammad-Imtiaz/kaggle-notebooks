```python
smx=0
sy=0
num=0
l=0
div=[]
conc=[]
im_height=range(0,3)
for p in digits.images:
    k=p[0:4]
    j=p[4:7]
    div.append(k)
    div.append(j)
    print("\n element \n",p," \n first part is \n",k,"\n")
    print("second part is : \n",j,"\n")
    
for m in div:
    for o in im_height:    
        for g in im_height:
            if m[o][g]==0: 
                smx=smx+o
                sy=sy+g
                num=num+1
    smx=smx/num
    sy=sy/num
    conc.append(smx)
    conc.append(sy)
    print("in image number ",l," X CENTER IS:(",smx,") AND  Y CENTER IS:(",sy,") \n")
    l=l+1
    
    
print("the centroid feature vector \n",conc)
digits.data=conc
  
for i in digits.images:
    k=i[0:4]
    j=i[4:7]
    figure,axes=plt.subplots(1,2,figsize=(8,4))
    for ax,image,number in zip(axes.ravel(),digits.images,y):
        ax.axis('off')
        ax.imshow(k,cmap=plt.cm.gray_r)
        ax.imshow(j,cmap=plt.cm.gray_r)
        ax.set_title('number' +str(number))        
    
```

```python
conc=x.tolist()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(conc,y,test_size=0.33,random_state=99)
```

```python
y_test.shape
```

```python
print(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y_pred)
print(report)
```

```python

from sklearn.metrics import confusion_matrix
confusion=confusion_matrix(y_true=y_test,y_pred=y_pred)
confusion
```

```python
import seaborn as sns
s=sns.heatmap(confusion,annot=True,cmap='nipy_spectral_r')
s.set_title('confusion matrix for MNIST dataset :')
None
```

```python
acc=knn.score(x_test,y_test)
print(f'accuracy from knn score={acc:0.4}')
from sklearn import metrics
acc=metrics.accuracy_score(y_test,y_pred)
print(f'accuracy from metrics.accuracy_score={acc:0.4}')
acc=metrics.balanced_accuracy_score(y_test,y_pred)
print(f'accuracy from balanced_accuracy_score ={acc:0.4}')
```
