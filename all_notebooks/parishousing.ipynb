```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/kaggle/input/paris-housing-price-prediction/ParisHousing.csv')
```

```python
df['hasYard'] = df['hasYard'].astype('bool')
df['hasPool'] = df['hasPool'].astype('bool')
df['isNewBuilt'] = df['isNewBuilt'].astype('bool')
df['hasStormProtector'] = df['hasStormProtector'].astype('bool')
df['hasStorageRoom'] = df['hasStorageRoom'].astype('bool')
df = df.drop(columns=['basement','attic','garage'])
```

```python
reg = RandomForestRegressor()
```

```python
X = df
y = X.pop('price')
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
reg.fit(X_train,y_train)
```

```python
reg.score(X_train,y_train)
```

```python
def evaluate(model, X, y):
    predictions = model.predict(X)
    errors = abs(predictions - y)
    mape = 100 * np.mean(errors / y)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    
    return accuracy
```

```python
evaluate(reg,X_train,y_train)
```

```python
evaluate(reg,X_test,y_test)
```

```python
feature_importance = reg.feature_importances_
```

```python
for i in range(0,len(feature_importance),1):
    new_array = zip(X.columns,feature_importance)
```
