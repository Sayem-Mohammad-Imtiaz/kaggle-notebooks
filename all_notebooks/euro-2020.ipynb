# Importing the libraries

```python
import plotly.graph_objects as go
from plotly.offline import iplot
```

# Loading the data

```python
data = pd.read_csv('/kaggle/input/euro-2020-top-players/euro2020.csv')
```

```python
data.head()
```

# greatest scorers

```python
dataDf2 = data.sort_values(by=['Goals'],ascending=False)
dataDf2 = dataDf2[dataDf2['Goals'] > 0]
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x= dataDf2['Player'],
    y= dataDf2['Goals'],
    name='Goals',
    hovertext = dataDf2['Country']
))
fig.add_trace(go.Bar(
    x= dataDf2['Player'],
    y= dataDf2['Total attempts'],
    name='Attempts',
    hovertext = dataDf2['Position']
))
fig.add_trace(go.Scatter(
    x= dataDf2['Player'],
    y= dataDf2['On target'],
    name='On target',
    hovertext = dataDf2['Position']
))
fig.add_trace(go.Scatter(
    x= dataDf2['Player'],
    y= dataDf2['Off target'],
    name='Off target',
    hovertext = dataDf2['Position']
))
fig.update_layout(title={
        'text': "Players with more goals",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'}, 
                  yaxis=dict(
                            title='goals',
                            titlefont_size=16,
                            tickfont_size=14),
                  xaxis=dict(
                            title='Player',
                            titlefont_size=16,
                            tickfont_size=14),
                  template = "plotly_dark"
                  )
iplot(fig)
```
