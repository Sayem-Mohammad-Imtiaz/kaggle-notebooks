```python
#clean
call_clean=call.copy()
```

```python
#tidness
call_clean=call_clean.drop('e',axis=1)
```

```python
list(call_clean)
```

```python
# Quality
call_clean['zip']=call_clean['zip'].fillna(call_clean['zip'].mean)
call_clean['twp']=call_clean['twp'].fillna(call_clean['twp'].mean)
```

```python
# Quality
call_clean.zip = call_clean.zip.astype(str)
call_clean['timeStamp'] = pd.to_datetime(call_clean['timeStamp'])
```

```python

call_clean.info()
```

```python
call_clean['title'].nunique()
```

```python
call_clean['title'].value_counts()
```

```python
 call_clean['type'] = call_clean['title'].apply(lambda x: re.search(r'^\w+', x).group(0))
```

```python
 call_clean['type']
```

```python
call_clean['type'].value_counts()
```

```python
call_clean=call_clean.drop('title',axis=1)
call_clean.head()
```

```python
#tidness
type(call_clean['timeStamp'].iloc[0])
time = call_clean['timeStamp'].iloc[0]
call_clean['Hour'] = call_clean['timeStamp'].apply(lambda time:time.hour)
call_clean['Month'] = call_clean['timeStamp'].apply(lambda time:time.month)
call_clean['DayofWeek'] = call_clean['timeStamp'].apply(lambda time:time.dayofweek)
call_clean['year'] = call_clean['timeStamp'].apply(lambda time:time.year)
call_clean['hour'] = call_clean['timeStamp'].apply(lambda x: x.hour)
call_clean['minute'] = call_clean['timeStamp'].apply(lambda x: x.minute)
call_clean['second'] = call_clean['timeStamp'].apply(lambda x: x.second)
```

```python
call_clean=call_clean.drop('timeStamp',axis=1)
```

```python
call_clean.head()
```

```python
#tidness
c=call_clean.drop('desc',axis=1)
```

```python
call_clean.head()
```

```python
def t_n(x):
    if x=='EMS': return 2
    if x=='Fire':   return 1
    if x=='Traffic':   return 0
```

```python
call_clean['type']=call_clean['type'].apply(t_n)
call_clean['type'].head
```

```python
call_clean=call_clean.drop('desc',axis=1)
```

```python
call_clean.head()
```

```python
call_clean['twp'] = call_clean['twp'].astype('str')
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
call_clean['twp']=le.fit_transform(call_clean['twp'])
call_clean['twp']
```

```python
call_clean['addr'] = call_clean['addr'].astype('str')
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
call_clean['addr']=le.fit_transform(call_clean['addr'])
call_clean['addr']
```

```python
call_clean['zip'] = call_clean['zip'].astype('str')
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
call_clean['zip']=le.fit_transform(call_clean['zip'])
call_clean['zip']
```

```python
call_clean=call_clean.drop('Hour',axis=1)
call_clean=call_clean.drop('minute',axis=1)
call_clean=call_clean.drop('second',axis=1)

```

```python
#classfication
from sklearn import tree
from sklearn.model_selection import train_test_split
X = nm.array(call_clean.drop(columns=['type']))
Y = nm.array(call_clean['type'])
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.33, random_state=42)
train_test_split(Y, shuffle=False)
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
y_pred
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,y_pred)
```
