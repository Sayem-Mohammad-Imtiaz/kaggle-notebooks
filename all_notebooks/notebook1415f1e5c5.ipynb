```python
import tensorflow as tf
import skimage
import os
#from tensorflow.keras import
import matplotlib.pyplot as plt
from PIL import  Image
import  numpy as np

rootpath='../input/artists-in-genres/genre'
images=[]
labels=[]
def load_data(rootpath):
    label_num=-1
    label2=[]
    for d in os.listdir(rootpath):
        label_path=rootpath+'/'+d
        label_num= label_num+1
        label2.append(label_num)
        for i in os.listdir(label_path)[:5]:
            image_path=label_path+'/'+i
            image_load=Image.open(image_path).resize((128,128),Image.BILINEAR)
            image_load = np.asarray(image_load)


            #如果不是三通道 则其他GB层生成0
            if (image_load.ndim==3):

                images.append(image_load)
            else:#如果图片不是三通道，改三通道再丢入数组
                image_load=np.expand_dims(image_load, axis=2)
                image_load = np.concatenate((image_load, image_load, image_load), axis=-1)
                #print(image_load.ndim)
                #print(image_load)
                #plt.imshow(image_load)
                #plt.show()
                images.append(image_load)
            labels.append(d)
            label2.append(label_num)
    return images,labels,label2

images,labels,label2=load_data(rootpath)

images=np.array(images)
label2=np.array(label2[0:65])

print(images.shape)
print(label2.shape)

#label2=label2.astype(np.int64)
#print(images.shape)

plt.imshow(images[20])
plt.show()


```

```python
import tensorflow as tf
import skimage
import os
from tensorflow import keras
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import tensorflow as tf
#import dataload
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout, Conv2D,MaxPool2D,AvgPool2D\
    ,Flatten
```

```python
print(images.shape)
print(label2.shape)
```

```python


model=keras.models.Sequential()
model.add(Conv2D(filters=32, kernel_size=(5,5),activation='relu',input_shape=(128,128,3)))
model.add(MaxPool2D(2,2))

model.add(Conv2D(filters=64,kernel_size=(5,5),padding='same',activation='relu'))
model.add(MaxPool2D(2,2))

model.add(Conv2D(filters=128,kernel_size=(3,3),padding='same',activation='relu'))
model.add(MaxPool2D(2, 2))

model.add(Conv2D(filters=128,kernel_size=(3,3),padding='same',activation='relu'))
model.add(MaxPool2D(2, 2))

model.add(Flatten())

model.add(Dense(1024,activation='relu', activity_regularizer=keras.regularizers.l1(0.01)))

model.add(Dense(512, activation='relu', activity_regularizer=keras.regularizers.l1(0.01)))
model.add(Dense(64, activation='relu', activity_regularizer=keras.regularizers.l1(0.01)))
model.add(Dense(13, activation='softmax'))

#model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=["accuracy"])

#print(tf.__version__)

model.compile(optimizer='adam',
              loss=keras.losses.categorical_crossentropy(from_logits=False),
              metrics=['accuracy']
              )
model.fit(images,label2,batch_size=20,epoches=5)


model.summary()
```
