Neuroticism:
Individuals who score high on neuroticism are more likely than average to be moody and to experience such feelings as anxiety, worry, fear, anger, frustration, envy, jealousy, guilt, depressed mood, and loneliness.

Extraversion:
It indicates how outgoing and social a person is. A person who scores high in extraversion on a personality test is the life of the party. They enjoy being with people, participating in social gatherings, and are full of energy.

Agreeableness:
When it comes to personality traits, if you're someone that has scored high in agreeableness, you're likely popular and tend to make friends easily.

Conscientiousness:
The definition of conscientiousness is a trait where you act based on what you know is right and with great care and honesty

Openess to experience:
1)People who tend to be high in the trait of openness are more willing to embrace new things, fresh ideas, and novel experiences. They are open-minded and approach new things with curiosity and tend to seek out novelty.
2)People who are low in openness to experience are typically very dedicated to any work they do, and always make sure to see their tasks through to the end. You are not very artistic or imaginative, and do not see the value in unnecessary things like art and music.
