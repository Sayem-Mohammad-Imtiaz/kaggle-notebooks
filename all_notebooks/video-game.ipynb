# **Now we will start rewiewing video games by year**

```python
total_column = "Total_Shipped"
df_ga[total_column] = df_ga["Global_Sales"]
df_ga.Year.value_counts(dropna = True)
mean_of_year= 2006.406443
```

```python
copy = df_ga.copy()
copy.Year= df_ga.Year.replace(["empty"],mean_of_year)
copy.Year.value_counts(dropna = True)
```

```python
copy = copy.sort_values("Year", ascending = True)
```

```python
copy.Year.value_counts(dropna = True)
```

```python
ani= px.histogram(copy, x= "Publisher", animation_frame= "Year", range_y =(0,70))
ani.update_xaxes(type= "category")
ani.update_xaxes(categoryorder = "category ascending")
ani.show()
```

```python
top_sal = copy.groupby(["Publisher","Year"]).agg({total_column: "count"}).reset_index()
top_sal
```

```python
top_sal.columns = ["Publisher","Year","Count"]
top_sal = top_sal[top_sal.Year.isin([2016,2017,2018,2019])]
top_sal = top_sal
top_sal = top_sal[top_sal.Count > top_sal.Count.sum() * 0.01]
top_sal["Year"]= top_sal["Year"].astype(str)
```

```python
an_sal = copy.groupby(["Publisher","Year"]).agg({total_column: "count"}).reset_index()
an_sal.columns = ["Publisher","Year","Count"]
an_sal = an_sal[an_sal.Year.isin([2016,2017,2018,2019])]
an_sal.Year.value_counts()
```

```python
top_sal
```

```python
ani2 =px.bar(top_sal , x = "Publisher", y = "Count", color = "Year" , barmode = "group")
ani2.update_layout(title = "Total Sales by Publishers")
ani2.update_xaxes(type = "category")
ani2.update_xaxes(categoryorder = "category ascending")
```

```python
ani3 =px.bar(an_sal , x = "Publisher", y = "Count", color = "Year" , barmode = "group")
ani3.update_layout(title = "Total Sales by Publishers")
ani3.update_xaxes(type = "category")
ani3.update_xaxes(categoryorder = "category ascending")
```
