```python
# lots of columns have lot of null values so its better to drop them 
df.drop(columns=['Education Place' ,'Native Language' , 'Mother' ,'Spouse', 'Father'] , axis =1 , inplace = True)
```

```python
import datetime
```

```python
df['Birth Date'] = pd.to_datetime(df['Birth Date'], errors = 'coerce')
df['Death Date'] = pd.to_datetime(df['Death Date'], errors = 'coerce')

# df['Age'] = round(df['Age']/np.timedelta64(1, 'Y'))
```

```python
'''np.where(condition,[x,y])
so in the below code when the Death Date column has null value then it means he is alive or the value is not provided
if he is alive then we will take the present data and subtract it with the Birth Data provided to get the number of days of age
else if Death Date is Not Null then we will calculate the no of days she was alive'''

df['Age'] = np.where(df['Death Date'].isna(),
                     datetime.datetime.now().date() - df['Birth Date'].dt.date, 
                     df['Death Date'].dt.date - df['Birth Date'].dt.date)
```

```python
df['Age'] = round(df['Age']/np.timedelta64(1, 'Y')) # Y here represents that the age will be shown in Years 
```

# Women Job Roles based on Description

```python
#This uses The Counter function from the collections module
# Easy method of plotting the same graph can be seen in the next cell

dict_ = Counter(df['Description'].value_counts().to_dict())
dict_ = dict(dict_.most_common(5))

plt.figure(figsize=(8,8))
plt.pie(x=dict_.values(), labels=dict_.keys(), autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```

```python
c = df['Description'].value_counts()
plt.figure(figsize=(12,8))
plt.pie(x=c.values[:5] , labels=c.keys()[:5], shadow=True , autopct='%.1f%%', startangle=12)
plt.title("Women in Different Roles based on Description")

plt.show()
```
