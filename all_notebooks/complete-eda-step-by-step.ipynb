```python
df = pd.read_csv('/kaggle/input/7k-indian-famous-women/data.csv')
df.head()
```

```python
df.shape
```

```python
df.isna().sum()  #we have quite a few null values to deal with
```

```python
df[['Description','Job']].sample(10)
#it seems like both the columns convey the same thing. It is safe to drop Description column
#actually no we can fill the missing values of job column with description since ** THEYRE MORE OR LESS THE SAME **
```

```python
df[df['Job'].isna()].head()  #observe 46th data 
```

```python
df['Job'] = df['Job'].fillna(df['Description'])
```

```python
df.iloc[46]  # here we can see that our values have been updated, awesome
```

```python
df.isna().sum()  #the null values in Job column have dropped down drastically, yaaay
```

```python
df = df.drop(['Description'],axis=1)  #dropping the column we dont need
```

```python
df['Job'].value_counts()  # we have 423 job categories they all fall under
```
