**Read all the csv files into Pandas DataFrames and Concat them all together into one DataFrame**

```python
data_2004 = pd.read_csv('/kaggle/input/lok-sabha-election-candidate-list-2004-to-2019/LokSabha2004.csv')
data_2004.head()
```

```python
data_2009 = pd.read_csv('/kaggle/input/lok-sabha-election-candidate-list-2004-to-2019/LokSabha2009.csv')
data_2014 = pd.read_csv('/kaggle/input/lok-sabha-election-candidate-list-2004-to-2019/LokSabha2014.csv')
data_2019 = pd.read_csv('/kaggle/input/lok-sabha-election-candidate-list-2004-to-2019/LokSabha2019.csv')
```

```python
data_2004['year'] = 2004
data_2009['year'] = 2009
data_2014['year'] = 2014
data_2019['year'] = 2019
```

***Concat method to concat the frames***

```python
data = pd.concat([data_2004, data_2009, data_2014, data_2019])
```

>To get the Candidate with most number of Criminal Cases

> City with highest Criminal Cases
* To make things easier lets create a new DataFrame with only "Criminal Cases" and "City" features

```python
crimes_city = pd.DataFrame()
crimes_city['Criminal Cases'] = data['Criminal Cases']
crimes_city['City'] = data['City']
```

```python
crimes_city.head()
```

* Group all rows based on their 'City' by adding up values of other columns
* Sort the same on number of Criminal Cases in descending order to get cities with highest crimes

```python
top_cities = crimes_city.groupby('City').sum().sort_values(by='Criminal Cases', ascending=False)[:10]
```

```python
top_cities.plot(kind='bar')
```

* Bar plot representing top 10 cities with highest crime cases
