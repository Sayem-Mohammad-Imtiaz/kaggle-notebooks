```python
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv('../input/carscsv/cars.csv')
```

```python
dataset
```

```python
x = dataset.iloc[:, 0:5].values
x
```

```python
y = dataset.iloc[:, 5].values
y
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x = sc_x.fit_transform(x)
x
```

```python
sc_y = StandardScaler()
y = sc_y.fit_transform(y.reshape(y.size, 1))
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
```

```python
classifier.add(Dense(units = 12, kernel_initializer = 'normal', activation = 'relu', input_dim = 5)) 
classifier.add(Dense(units = 12, kernel_initializer = 'normal', activation = 'relu')) 
classifier.add(Dense(units = 1, kernel_initializer = 'normal', activation = 'linear'))
classifier.summary()
```

```python
classifier.compile(optimizer='adam', loss='mean_squared_error', metrics = ['mae', 'mse'])
```

```python
classifier.fit(x_train, y_train, batch_size = 10, epochs = 50, shuffle = True)
```

```python
y_pred = classifier.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
rSquare = r2_score(y_test, y_pred)
```
