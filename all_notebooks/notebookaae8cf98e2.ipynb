# Import File

```python
df = pd.read_csv("/kaggle/input//pubg-weapon-stats/pubg-weapon-stats.csv")
```

# Check the first 3 items from head

```python
print(df.head(3))
```

# Check columns and data types

```python
print(df.dtypes)
```

# Check Null in this DataFrame

```python
if not df.isnull().values.any():
    print("\nAll rows are filled")
else:
    print("\nNull Exists: Displaying The Rows:")    
    print(df[df.isnull().any(axis=1)]) # displaying specific rows where contains NUll
```

# Fill with 0
**From the above results, we can fill with 0.**

```python
df = df.fillna(0)
```

# Check summary based on Weapon Type

```python
print(df.groupby('Weapon Type').describe())
```

# Plotting range(numerical) and damage(numerical) by weapon types
# Showing the Correlation
(+) 1 has positive relations, on the other hand (-) 1 is negative relations

```python
grouped = df.groupby('Weapon Type')
for name, group in grouped:
    group.plot.scatter(x="Range", y="Damage")
    plt.title(str(name))
    plt.grid(True)
plt.show()

print(df.groupby('Weapon Type')[['Range', 'Damage']].corr())
```

# Plotting Shots to Kill (Chest)/(Head) and damage(numerical) by weapon types
# Showing the Correlation

```python
# Chest 

for name, group in grouped:
    group.plot.scatter(x="Shots to Kill (Chest)", y="Damage")
    plt.title(str(name))
    plt.grid(True)
plt.show()

print(df.groupby('Weapon Type')[['Shots to Kill (Chest)', 'Damage']].corr())


# Head

for name, group in grouped:
    group.plot.scatter(x="Shots to Kill (Head)", y="Damage")
    plt.title(str(name))
    plt.grid(True)
plt.show()

print(df.groupby('Weapon Type')[['Shots to Kill (Head)', 'Damage']].corr())
```

# Plotting damage and Shots to Kill (Chest)/(Head) relations

```python
plt.subplot(1, 2, 1)
plt.scatter(df['Shots to Kill (Chest)'], df['Damage'])
plt.ylabel("Damage")
plt.title('Shots to Kill (Chest)')

plt.subplot(1, 2, 2)
plt.scatter(df['Shots to Kill (Head)'], df['Damage'])
plt.title('Shots to Kill (Head)')

plt.show()
```

# Plotting damage and Shots to Magazine Capacity relations

```python
plt.scatter(df['Magazine Capacity'], df['Damage'])
plt.title('Magazine Capacity')
plt.ylabel("Damage")
plt.show()

print("\nAverage of Magazine Capacity and Damage by Weapon Types")
print(df.groupby('Weapon Type')[['Magazine Capacity', 'Damage']].mean())
```
