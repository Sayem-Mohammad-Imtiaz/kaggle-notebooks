```python
# Name: Kyle Cree
# Id: 1505970


import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib as plt
import seaborn as sns
```

```python
# Get the data and store in a panda dataframe
data = pd.read_csv("/kaggle/input/wisconsin-breast-cancer-cytology-features/wisconsin_breast_cancer.csv")

# replace anythings that is nan with 0
data.fillna(0, inplace=True)

# Display a pairplot with seabourn
sns.pairplot(data)
```

```python
# Look at some of the data
data.head(20)
```

```python
# Look at the data and what types they are as well as other information
data.info()
```

```python
# Import the train test split 
from sklearn.model_selection import train_test_split
# Create the X variable which is everything except the class column
X = data.drop("class", axis=1)
# Make the y varibale which is the class column
y = data['class']
# Split the X and y values into a training set and a test set, 20% test size and random_state=1505970
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1505970)
```

```python
# Import the SVC module
from sklearn.svm import SVC
# Leaving the SVC module blank defaults the kernal to rbf
model = SVC()
```

```python
# Fit the model to the training dataset
model.fit(X_train, y_train)
```

```python
# Create the predictions of the y values given the X_test values
pred = model.predict(X_test)
```

```python
# Import that classification and confusion matrix
from sklearn.metrics import classification_report, confusion_matrix
# Print the confustion matrix
print(confusion_matrix(y_test,pred))
# Print the classification report
print(classification_report(y_test,pred))
```

**Results**
Overal we find that the results are not good as they are heavily skewed to 0 in this case as there are no 1 results guessed.
The overal percision of the model is not good with an average of 43% correct, 66% correct on the 0 guess and none corect on the 1 guesses as no 1's were selected. Therefore creating a divide by zero problem that could throw that overal dataset values out even more.
Recall and f-1 have the same type of problem as there were no 1's predicted and therefore the divide by zero problem is still there

This overal means that this model is not good, the model always predicts the values is being a 0 and therefore does not see any pattern between the rest of the data and the class. 

I have however found a way to improve on the basic SVC model and that is to use a different kernel type, in this case the best is the sigmoid function

```python
# Import the SVC module
from sklearn.svm import SVC
# Leaving the SVC module blank defaults the kernal to rbf
model = SVC(kernel='sigmoid')
# Fit the model to the training dataset
model.fit(X_train, y_train)
# Create the predictions of the y values given the X_test values
pred = model.predict(X_test)
# Import that classification and confusion matrix
from sklearn.metrics import classification_report, confusion_matrix
# Print the confustion matrix
print(confusion_matrix(y_test,pred))
# Print the classification report
print(classification_report(y_test,pred))
```

**Result**
When using the sigmoid data the model preforms better, the main difference is the model predicts the values of 1's instead of just 0's. Therefore improving all the scores and making the data not skewed as there is not a divide by zero problem.

Overal the weighted averages for all of the values are better then the previous models, however they are not good enough to be used in the medical world as the predictions are not good enough.

There are ways to improve the model further, the main on is to use more data, as machine learning benifits from the use of more data to improve the model. As well as this other methods of machine learning such as K-nearest neightbour could be applied and see if the results are better.
