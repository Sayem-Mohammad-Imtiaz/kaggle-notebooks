```python
df = pd.read_csv('../input/goodreadsbooks/books.csv',error_bad_lines=False)
```

```python
df.head(5)
```

```python
most_reviewed = df.sort_values(by='text_reviews_count', ascending=False) # Sorting by number of reviews
most_reviewed = most_reviewed[['title','text_reviews_count']] # Selecting just the columns we want
most_reviewed = most_reviewed.head(10) # Choosing only the top 10 books
```

```python
# Selecting the first book of this dataset
Best_Title, Number_votes = (most_reviewed['title'].iloc[0], most_reviewed['text_reviews_count'].iloc[0])
print(f'{Best_Title} is the most reviewed book with a total of {Number_votes} votes')
```

```python
# Plotting the top 10 books with highest quantity of reviews
plt.figure(figsize=(12,5))
sns.barplot(y='title', x='text_reviews_count', data = most_reviewed, palette="rocket")
plt.title('Most Reviewed Books', fontsize='20')
plt.xlabel('Reviews Count')
plt.ylabel('Title')

for index,value in enumerate(most_reviewed['text_reviews_count']): #Adding the values to the graph
    plt.text(value, index, str(value)).set_fontsize(9)
plt.show()
```
