```python
import pandas as pd
import numpy as np
import collections
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("/kaggle/input/netflix-shows/netflix_titles.csv")
```

```python
# Number of TV Shows vs Movies
count_movies = len(dataset[dataset['type'] == 'Movie'])
count_tv_show = len(dataset[dataset['type'] == 'TV Show'])
print("Count of Movies = %d, Count of TV shows = %d" % (count_movies, count_tv_show))
```

```python
# Top 10 Directors whom Netflix really likes ...
dataset['director'].value_counts()[:10]
```

```python
# Most recurring actors in this dataset including both Movies and TV Shows
def count_cast(list_cast):
    list_all = []
    for names in list_cast:
        try: # split throws error on nan values
            for name in names.split(","):
                list_all.append(name)
        except:
            continue
    counter = collections.Counter(list_all)
    print(counter.most_common(10))
list_cast = dataset['cast']
count_cast(list_cast)
```

```python
# Top 10 common recurring cast on movies
list_cast = dataset[dataset['type'] == 'Movie']['cast']
count_cast(list_cast)
```

```python
# Top 10 common recurring cast on TV Shows
list_cast = dataset[dataset['type'] == 'TV Show']['cast']
count_cast(list_cast)
```
