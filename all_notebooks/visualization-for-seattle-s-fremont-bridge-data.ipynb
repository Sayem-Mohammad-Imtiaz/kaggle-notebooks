Below data comes from an automated bicycle counter, installed in late 2012, which has inductive sensors on the east and west sidewalks of the bridge. 

```python
import pandas as pd
data = pd.read_csv("../input/fremont-bridge/fremont-bridge.csv", index_col= 'Date', parse_dates=['Date'])
data.head()
```

```python
data.columns = ["West", "East"]
data["Total"] = data["West"] + data["East"] 
data.head()
```

```python
data.info()
```

# Visualizing the data

```python
import matplotlib.pyplot as plt
data.plot(figsize=(20, 4))
plt.grid()
plt.legend(loc='best')
plt.ylabel("Hourly Bicycle count")
plt.show()
```

```python
data.isnull().sum()* 100 / len(data)
```

Since, Missing value is quite less. So, we will drop NA

```python
data.dropna().describe()
```

```python

import seaborn
seaborn.set()
data.plot(figsize=(20,5))
plt.ylabel("Hourly Bicycle count")
plt.show()
```

```python
yearly = data.resample("Y").sum()
yearly.plot(style=[':', '--', '-'],figsize=(20,5))
plt.ylabel('Yearly bicycle count')
plt.show()
```

```python
weekly = data.resample("W").sum()
weekly.plot(style=[':', '--', '-'],figsize=(20,5))
plt.ylabel('Weekly bicycle count')
plt.show()
```

```python
daily = data.resample('D').sum()
daily.rolling(30, center=True).sum().plot(style=[':', '--', '-'],figsize=(20,5))
plt.ylabel('Mean Hourly Count')
plt.show()
```

```python
daily.rolling(50, center=True,
              win_type='gaussian').sum(std=10).plot(style=[':','--', '-'],figsize=(20,5))
plt.ylabel('Mean Hourly Count')
plt.show()
```

```python
import numpy as np
by_time = data.groupby(data.index.time).mean()
hourly_ticks = 4 * 60 * 60 * np.arange(6)
by_time.plot(xticks= hourly_ticks, style=[':', '--', '-'],figsize=(20,5))
plt.ylabel("Traffic according to Time")
plt.show()
```
