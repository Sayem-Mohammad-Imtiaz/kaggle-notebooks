What determines a Pokemon's type? We have an idea of how strong they are based on certain stats, but what if a new Pokemon is discovered, what type (or types) should it be designated. These are the hard questions Professor Oak and others like him have to deal with with new discoveries. 

An overview of the data for those who aren't as familiar with the games (i.e. me):

* \# : ID for each pokemon
* Name: Name of each pokemon
* Type 1: Each pokemon has a type, this determines weakness/resistance to attacks
* Type 2: Some pokemon are dual type and have 2
* Total: sum of all stats that come after this, a general guide to how strong a pokemon is
* HP: hit points, or health, defines how much damage a pokemon can withstand before fainting
* Attack: the base modifier for normal attacks (eg. Scratch, Punch)
* Defense: the base damage resistance against normal attacks
* SP Atk: special attack, the base modifier for special attacks (e.g. fire blast, bubble beam)
* SP Def: the base damage resistance against special attacks
* Speed: determines which pokemon attacks first each round
