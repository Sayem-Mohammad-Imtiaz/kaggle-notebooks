Dataset bao gồm số báo danh thí sinh và điểm các tương ứng của thí sinh đó:

Ma Ngoai Ngu - Mã môn ngoại ngữ/Foreign language code

Ngoai Ngu - Điểm ngoại ngữ/Foreign language score

N1: Tiếng Anh/English

N2: Tiếng Nga/Russian

N3: Tiếng Pháp/French

N4: Tiếng Trung/Chinese

N5: Tiếng Đức/German

N6: Tiếng Nhật/Japanese

Vat Ly - Điểm vật lý/Physics score

Hoa Hoc - Điểm hóa học/Chemistry score

Sinh Hoc - Điểm sinh học/Biology score

Lich Su - Điểm lịch sử/History score

Dia Ly - Điểm địa lý/Geography score

GDCD - Điểm giáo dục công dân/Civics score
