```python
from dask import dataframe as dd
import dask
```

First, let's load the data with Dask.

```python
df = dd.read_csv('/kaggle/input/medal-emnlp/full_data.csv')
df
```

Let's take a look at what the data looks like:

```python
df.head()
```

Now, let's apply a mask using the word "substrate", and create a smaller dataframe:

```python
mask = df.LABEL.str.contains('substrate', regex=False)
substrate = df[mask]
substrate
```

Notice that the dataframe itself is delayed:

```python
substrate.shape
```

To get the entire dataframe, we need to compute it:

```python
%%time
computed = dask.compute(substrate, scheduler='processes')[0]
computed.head()
```

Let's see if it worked:

```python
print(computed.shape)
print(type(computed))
```

Awesome! Now you can use it the usual (pandas) way.
