```python
#What if rubik competition is like kaggle competition?

#Which country mostly get gold coin? silver coin? bronze coin?

#We will answer these question with this notebook.
```

```python
#import necessary modules
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('../input/evolution-of-rubiks-cube-solve-times/best_singles.csv')
#drop unuse columns
data.drop(['name','competition','Unnamed: 0','time'],axis=1,inplace=True)

#check number of competitors
len(data)
```

```python
# Since there is 1000 competitor
# top 10 will get gold medal
# top 5% will get silver medal
# top 10% will get bronze medal

# Therefore rank 1-10 will get gold coin, rank 11-50 will get silver coin, rank 50-100 will get bronze coin

# define medal check function
def medal_func(row):
    if row['rank']<=10:
        return 'gold'
    elif 10<row['rank']<=50:
        return 'silver'
    elif 50<row['rank']<=100:
        return 'bronze'
    else:
        return '-'

#apply function to dataframe with column axis
data['medal']=data.apply(medal_func,axis=1)
data
```

```python
#convert medal column into dummies
medal_dummies=pd.get_dummies(data['medal'])

#drop '-' column
medal_dummies=medal_dummies.drop(['-'],axis=1)
medal_dummies
```

```python
#concatenate with main dataframe with column axis
data=pd.concat([data,medal_dummies],axis=1)

#drop medal column
data.drop('medal',axis=1,inplace=True)

data
```

```python
#group data with country and sum it
medal_data=data.groupby('country').sum()

#drop rank column
medal_data.drop('rank',axis=1,inplace=True)

#(optinal) rearrage column
medal_data=medal_data.loc[:,['gold','silver','bronze']]
medal_data
```

```python
#top country that got most gold coin
medal_data.sort_values('gold',ascending=False).head()
```

```python
#top country that got most silver coin
medal_data.sort_values('silver',ascending=False).head()
```

```python
#top country that got most bronze coin
medal_data.sort_values('bronze',ascending=False).head()
```

```python
#we also want to know which country get most medal

#sum medal and store it in new column
medal_data['total']=medal_data['gold']+medal_data['silver']+medal_data['bronze']
medal_data
```

```python
#top country that got most coin
medal_data.sort_values('total',ascending=False).head()
```

```python
#we can also plot bar chart for better compare medal of each country

#before we plot bar chart we should drop country with 0 coin

#find country that get 0 coin
zero_coin=medal_data.loc[medal_data.total==0,:]

zero_coin.index
```

```python
#drop it
medal_data_for_plot=medal_data.drop(zero_coin.index)
medal_data_for_plot
```

```python
#plot bar chart

#set fig size
fig=plt.figure(figsize=(32,8))

#bronze coin
#set location, height, color, bar width, label
plt.bar(np.arange(len(medal_data_for_plot)),medal_data_for_plot['bronze'],color='#CD7F32',width=0.5,label='bronze')

#silver coin
#stack silver coin with bronze coin, other parameter remain same
plt.bar(np.arange(len(medal_data_for_plot)),medal_data_for_plot['silver'],color='#C0C0C0',width=0.5,label='silver',bottom=medal_data_for_plot['bronze'])

#gold coin
#stack gold coin with silver and bronze coin, other parameter remain same
plt.bar(np.arange(len(medal_data_for_plot)),medal_data_for_plot['gold'],color='#FFD700',width=0.5,label='gold',bottom=medal_data_for_plot['bronze']+medal_data_for_plot['silver'])

#next, we will chance xtick names

#before change xtick name, strip and shorten some country's name
country_tick=[country.strip(' ') for country in medal_data_for_plot.index]

country_tick[4]='Czech'
country_tick[20]='Korea'
country_tick[28]='UK'
country_tick[29]='USA'

plt.xticks(np.arange(len(medal_data_for_plot)),country_tick,fontsize=10);
```
