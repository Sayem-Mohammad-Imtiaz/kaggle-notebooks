```python

#IMPORT NECESSARY LIBRARIES 

import cv2  #
import pandas as pd 

# --------------------------------------------------------------------------
# Taking an image from the user
img_path = '/content/colorpic.jpg'
csv_path = '/content/colors.csv'

# Reading csv file with pandas and giving names to each column
index = ['color', 'color_name', 'hex', 'R', 'G', 'B']
df = pd.read_csv(csv_path, names=index, header=None)

# reading image with opencv
img = cv2.imread(img_path)
img = cv2.resize(img, (800,600))

# it means that we have not double clicked left button until now 
clicked= False
r=g=b= xpos = ypos = 0

def get_color_name(R,G,B):
    minnimum=1000
    for i in range(len(df)):
        d = abs(R-int(df.loc[i, 'R'])) +  abs(G-int(df.loc[i, 'G'])) + abs(B-int(df.loc[i, 'B'])) 
        if d<= minimum:
            minimum = d
            cname =df.loc[i,'color_name']
            return cname
        
# it tells about the x , y coordinate position of the cursor
def draw_function(event,x,y,flags,params):
    
    # it check e=whether we have clicked lef button or not
    if event== cv2.EVENT_LBUTTONDBLCLK: 
        
        # this function is used to make change appear outside the function by making variable global
        global clicked,r,g,b,xpos,ypos
        
        clicked=True
        xpos = x
        ypos = y
        b,g,r =img[y,x]
        b=int(b)
        g=int(g)
        r=int(r)

# to display image
cv2.namedWindow('image')

cv2.setMousseCallback('image', draw_function)

# it is used so that our program keep on running
while True:
    cv2.imshow('image',img)
    if clicked:
        # this coding is used to make rectangle in which our colour name is showed
        cv2.rectangle(img,(20,20),(600,600),(b,g,r),-1)
        text = get_color_name(r,g,b) + ' R=' + str(r) + ' G=' + str(g) + ' B=' + str(b)
        
        # to put text in rectangle
        cv2.putText(img,text,(50,50),2,0.8,(255,255,255),2,cv2.LINE_AA)
        
        # if colour is light 
        if r+g+b >=600:
        cv2.putText(img,text,(50,50),2,0.8,(0,0,0),2,cv2.LINE_AA)
        
     # if after 20 seconds there is no double click than it close the window
    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows() 
```
