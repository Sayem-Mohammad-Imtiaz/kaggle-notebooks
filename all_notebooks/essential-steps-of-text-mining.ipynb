
### Uppercase - Lowercase.

When we need to standardize the shape of the letter, we can use this method with ease.

```python
data1 = data.copy()
```

```python
# For Lowercase
data1.Message.apply(lambda i: " ".join(i.lower() for i in i.split()))
```

```python
#For Uppercase
data1.Message.apply(lambda i: " ".join(i.upper() for i in i.split()))
```

I want to keep the lowercase style of Data 1, then I can move on to the next step where I will clear the punctuation marks.

```python
data2 = data1.Message.apply(lambda i: " ".join(i.lower() for i in i.split()))
```

### Remove Punctuation Marks.

```python
data2.str.replace("[^\w\s]","")
```

```python
data3 = data2.str.replace("[^\w\s]","")
```

### Removing the words that its quantity lower and higher than normal.

**Before the process, I have to create a dataframe again to use some methods of pandas library.**

```python
data5
```

```python
data5 = pd.DataFrame(data5, columns=["Message"])
data5.head()
```

```python
#Looking at how many quantities each word there are in the dataset.
words_counts = pd.Series(" ".join(data5.Message).split()).value_counts()
words_counts
```

```python
low = words_counts[words_counts<5]
len(low)
```

```python
high = words_counts[words_counts>100]
len(high)
```

**If you think how you specify high and low border, it is completely secret in my heart :) I specified borders without analysis but if I chose an easy way to think the rule is so simple.  A high amount cause bias on the other hand low amount keep no insight for algorithms**

```python
data5 = data5.Message.apply(lambda x: " ".join(x for x in x.split() if x not in low))
data5 = data5.apply(lambda x: " ".join(x for x in x.split() if x not in high))
```

### Stemming

```python
from nltk.stem import PorterStemmer
st = PorterStemmer()
```

```python
data5.apply(lambda x: " ".join(st.stem(i) for i in x.split()))
```


So to speak stemming destroyed all of my words in data so I will apply lemmatization instead.

### Lemmatization

```python
from textblob import Word
nltk.download("wordnet")
```

```python
data5.apply(lambda x: " ".join(Word(i).lemmatize() for i in x.split()))
```

**At least lemmatization looks better than stemming.**
