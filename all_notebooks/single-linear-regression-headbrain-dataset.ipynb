```python
X = dataset.iloc[: ,2].values   #Features
y = dataset.iloc[: ,3].values   #Targets
```

```python
# plot the imput data
# -------------------


plt.scatter(X,y, color = 'purple', label = 'Data Points')
plt.xlabel('Head size in cm3')
plt.ylabel('Brain Weight weight in grams')
plt.legend()


plt.show()
```

```python
# from sklearn.linear_model import LinearRegression
```

```python
# Simple Linear Regression using Sklearn
# --------------------------------------

model = LinearRegression()
```

##### Link: https://scikit-learn.org/stable/modules/linear_model.html#ordinary-least-squares

```python
X.shape
```

```python
X = X.reshape(len(X), 1)     # We cannot use rank 1 matrices in sklearn so we have to reshape
```

```python
X.shape
```

```python
# model Fting 
# -----------

model.fit(X,y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
# Model Prediction
# ----------------

y_pred = model.predict(X)
```

```python
# y_pred
```

```python
# Visualising: Plotting Values and Regression Line
# -------------------------------------------------
# Ploting Values 
plt.scatter(X, y, color='green', label='Scatter Points')


# Ploting Regression Line 
plt.plot(X, y_pred,  color = 'blue', label = 'Regression Kine')

plt.title('Head Size VS Brain Weight')
plt.xlabel('Head Size(cm^3)')
plt.ylabel('Bran Weight(grams)')
plt.legend()
plt.show()
```

#### Râ€“squared value is the statistical measure to show how close the data are to the fitted regression line.

```python
model.score(X,y)
```

```python
# from sklearn.metrics import r2_score
```

```python
# Model Evaluation
# ................

r_squared_score = r2_score(y, y_pred)   # R2 (Coefficient of Sata Matrics) regression score function
```

```python
print((r_squared_score * 100).round(2), "%")
```
