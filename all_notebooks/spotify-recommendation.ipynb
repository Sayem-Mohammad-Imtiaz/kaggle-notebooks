```python
# Predictor dan target
x = df.drop('liked', axis = 1)
y = df['liked']
```

```python
x.head()
```

```python
y.head()
```

```python
# Splitting train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=32)
# Print according to the expected result
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc_model = GradientBoostingClassifier().fit(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
y_pred_gbc = gbc_model.predict(x_test)
accuracy_score(y_test, y_pred_gbc)
```

```python
# Print classification report
print('Classification Report (Gradient Boosting Classifier) :')
print(classification_report(y_test, y_pred_gbc))
```

```python
#evaluasi model dengan confusion matrix
cnf_matrix=confusion_matrix(y_test,y_pred_gbc)
print('Confusion Matrix\n',cnf_matrix)
```

```python
plt.figure()
heatmap = sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, annot_kws={'size': 14}, fmt='d', cmap='YlGnBu')
heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=14)
heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=0, ha='right', fontsize=14)

plt.title('Heatmap of Confusion Matrix\n(GradientBoostingClassifier)\n', fontsize=18)
plt.show()
```
