```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.sentiment import SentimentIntensityAnalyzer
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
review = pd.read_csv('../input/reviewuniversalstudio/universal_studio_branches.csv')
print(review.head(2))
```

```python
text = review.title.values
text = pd.DataFrame(text)
text = text.stack().str.replace("'", "").unstack()
text = text.stack().str.lower().unstack()
wordcloud = WordCloud(
    width = 3000,
    height = 2000,
    background_color = 'black',
    stopwords = STOPWORDS).generate(str(text))
fig = plt.figure(
    figsize = (40, 30),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

```python
text = review.review_text.values
text = pd.DataFrame(text)
text = text.stack().str.replace("'", "").unstack()
text = text.stack().str.lower().unstack()
wordcloud = WordCloud(
    width = 3000,
    height = 2000,
    background_color = 'black',
    stopwords = STOPWORDS).generate(str(text))
fig = plt.figure(
    figsize = (40, 30),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

```python
def tokenization(words):
    w_new = []
    for w in words:  
        w_token = word_tokenize(w)
        if w_token != '':
            w_new.append(w_token)
    return w_new

lemmatizer = WordNetLemmatizer()
```

```python
lemmas=[]
tokens = tokenization(review['title'])
for x in tokens:
    for word in x:
     lemmas.append(lemmatizer.lemmatize(word, pos='v'))
```

```python
lemmas = review.review_text.values
lemmas = pd.DataFrame(text)
lemmas = text.stack().str.replace("'", "").unstack()
lemmas = text.stack().str.lower().unstack()
wordcloud = WordCloud(
    width = 3000,
    height = 2000,
    background_color = 'black',
    stopwords = STOPWORDS).generate(str(lemmas))
fig = plt.figure(
    figsize = (40, 30),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```

```python
sia = SentimentIntensityAnalyzer()
for x in review['title']:
 sent = sia.polarity_scores(x)
```

```python
def is_positive(sent):
    #True if title has positive compound sentiment, False otherwise.
    if sia.polarity_scores(sent)["compound"] > 0:
     return 'Positive'
    else:
     return 'Negative'

title_pos = []
for x in review['title']:
 title_pos.append(is_positive(x))
```

```python
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(10,10))
title_pos = pd.DataFrame(title_pos)
sns.countplot(x=title_pos.columns[0], data=title_pos)
plt.xlabel('Sentiment')
plt.show();
```

```python
plt.figure(figsize=(10,10))
label = ['Positive', 'Negative']
color = ['orange', 'blue']
plt.pie(title_pos.value_counts(), labels=label, autopct='%1.0f%%', colors=color)
plt.title('Sentiment')
plt.show();
```
