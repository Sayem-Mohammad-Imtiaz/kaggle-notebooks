 <h1 style="background-color:#D7BDE2;font-family:newtimeroman;font-size:200%;text-align:center;border-radius: 15px 50px;"> üèéÔ∏è Checking which teams have won the biggest amount of championships and points 
 üèéÔ∏è</h1><a id=1.1></a>
  
#### [Back to the top](#0)

### Lets focus only on the teams Grand Prix wins:

```python
# Let's delete all teams that have never win championship
gry=cons
gry['Position'] = pd.to_numeric(gry['Position'], errors='coerce')
gry = gry[gry.Position == 1]

team_win_c= gry.groupby('Team')['Team'].count()
team_win_c = pd.DataFrame(team_win_c)
team_win_c.columns = ['Wins']
team_win_c.reset_index(level=0, inplace=True)

team_win_c.sort_values(by=['Wins'], inplace=True, ascending=False)
team_win_c = team_win_c.head(10)
team_win_c = team_win_c[::-1]

fig = px.bar(team_win_c, x='Team', y='Wins',color='Wins',width=750, height=500)
fig.update_layout(title={'text': 'Teams with The Most Grand Prix Won','y':0.95,'x':0.5})
fig.show()
```

### Lets focus on the teams points scored:
<a id=1.2></a>
 
#### [Back to the top](#0)

```python
# Let's count all points won by each team
team_points=cons[["Points","Team"]].groupby('Team').sum()
team_points.sort_values(by=['Points'], inplace=True, ascending=False)
print(team_points)
```

```python
# Let's count all points won by each team


cons["Points"] = pd.to_numeric(cons["Points"])
cons.sort_values(by=['Points'], inplace=True, ascending=False)
dfp = cons.groupby(["Team"]).agg({"Points":"sum"}).loc[lambda d: d["Points"].gt(100)]
dfp =dfp.sort_values('Points', ascending=True)
px.bar(dfp, x=dfp.index, y="Points", color="Points")
```

### Lets see how many drivers from each team  got the podium
<a id=1.3></a>
 
#### [Back to the top](#0)

```python
# Let's count how many times a driver from a team standed on the podium
gry1=cons
team_points=gry1[["Points","Team"]].groupby('Team').sum()
team_points=gry1[gry1['Position'].isin([1,2,3])]
ax= px.treemap(team_points,path=["Team"],title="Driver from a team standed on the podium")
ax.show()
```

### Lets check how the teams did in the last 2020 season
<a id=1.4></a>
 
#### [Back to the top](#0)

```python
gry3=cons
gry3["Year"] = pd.to_numeric(cons["Year"])
gry3=gry3[gry3['Year'].isin([2020])]
gry3["Points"] = pd.to_numeric(gry3["Points"])
gry3.sort_values(by=['Points'], inplace=True, ascending=True)
dfp = gry3.groupby(["Team"]).agg({"Points":"sum"}).loc[lambda d: d["Points"].gt(100)]
dfp =dfp.sort_values('Points', ascending=True)
px.bar(dfp, x=dfp.index, y="Points", color="Points")
```

### Lets check top four constructors achievments during years
<a id=1.5></a>
 
#### [Back to the top](#0)

```python
dfer= cons[cons.Team == "Ferrari"]
dfmer= cons[cons.Team == "Mercedes"]
dfwil= cons[cons['Team'].str.contains("Williams")]
dfrb=cons[cons['Team'].str.contains("Red Bull")]
dfer =dfer.sort_values('Year', ascending=False)
dfmer =dfmer.sort_values('Year', ascending=False)
dfwil =dfwil.sort_values('Year', ascending=False)
dfrb =dfrb.sort_values('Year', ascending=False)
fig = make_subplots(rows=2, cols=2)
fig.add_trace(go.Scatter(x=dfer['Year'], y=dfer['Points'], name="Ferrari"),row=1, col=1)
fig.add_trace(go.Scatter(x=dfmer['Year'], y=dfmer['Points'], name="Mercedes"),row=1, col=2)
fig.add_trace(go.Scatter(x=dfwil['Year'], y=dfwil['Points'], name="Williams"),row=2, col=1)
fig.add_trace(go.Scatter(x=dfrb['Year'], y=dfrb['Points'], name="Red Bull"),row=2, col=2)
fig.update_layout(height=600, width=650, title_text="Top 4 constructors")
fig.show()
```

### Summary of the constructors analyse
<a id=1.6></a>

 ##### * The best team overall from the beginning of the F1 is Ferrari
 ##### * The second biggest amount has gathered Mercedes altough he started racing in 2010
 ##### * One of the best teams Williams from 2014 significantly drops down his acquirements
  ##### * Red Bull that is also not racing from long time ago is in top 4 points scorers overall
   ##### * We can see a huge points advantage in season 2020 from Mercedes
