```python
# Constant
PATH_GENRE = "/kaggle/input/dataset-of-songs-in-spotify/genres_v2.csv"
X_COL = "instrumentalness"
Y_COL = "danceability"
```

```python
# Packages 
import pandas as pd
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
```

```python
# Parameters
matplotlib.rcParams['figure.figsize'] = [10, 10]
```

```python
# Read data
df = pd.read_csv(PATH_GENRE,low_memory=False)
```

```python
# Basic matplotlib scatter plot
plt.scatter(df[X_COL], df[Y_COL])
plt.xlabel(X_COL)
plt.ylabel(Y_COL)
plt.show()
```

```python
# Matplotlib scatter plot with an alpha value
plt.scatter(df[X_COL], df[Y_COL], alpha=0.03)
plt.xlabel(X_COL)
plt.ylabel(Y_COL)
plt.show()
```

```python
# Matplotlib scatter plot with an alpha value
plt.scatter(df[X_COL], df[Y_COL], alpha=0.01)
plt.xlabel(X_COL)
plt.ylabel(Y_COL)
plt.show()
```
