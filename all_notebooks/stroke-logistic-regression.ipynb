```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegressionCV
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('../input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')
df =df.fillna(0)
```

```python
Cat_Values = ['gender','ever_married', 'work_type', 'Residence_type', 'smoking_status']

for cat in Cat_Values:
    print(cat,':',df[cat].unique())
    
df.groupby(by = ['gender','stroke']).count()
```

```python
df_dummy = pd.get_dummies(df, ['gender','ever_married','work_type', 'Residence_type','smoking_status'])
```

```python
Y = df_dummy['stroke']
X = df_dummy[['age','gender_Female', 'gender_Male',
       'gender_Other', 'work_type_Govt_job', 'work_type_Never_worked',
       'work_type_Private', 'work_type_Self-employed', 'work_type_children',
       'Residence_type_Rural', 'Residence_type_Urban',
       'smoking_status_Unknown', 'smoking_status_formerly smoked',
       'smoking_status_never smoked', 'smoking_status_smokes']]

X_Train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.75, random_state = 1896425)
```

```python
Stroke_Model = LogisticRegressionCV(cv = 5, random_state = 891532, solver = 'liblinear').fit(X_Train, Y_train)
print('Training Score: ', Stroke_Model.score(X_Train, Y_train))
print('Test Score: ',Stroke_Model.score(X_test,Y_test))
# clf.get_params(deep = False)
```
