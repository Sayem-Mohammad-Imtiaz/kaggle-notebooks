# Data Cleaning
* cahnge data type
* fill the null the mean for  each specific region

```python
df_19[df_19.columns[4:]]=df_19[df_19.columns[4:]].astype(float)
```

```python
count=0
while count<len(df_19.region.unique()):
    m=round(df_19[df_19['region']==df_19.region.unique()[count]].mean(),2)
    (df_19[df_19['region']==df_19.region.unique()[count]])=(df_19[df_19['region']==df_19.region.unique()[count]].fillna(m))
    count+=1
    
```

```python
df_19.isnull().sum()
```

```python
df_19["hf_rank"]=df_19.hf_rank.astype(int)
```
