# **t-SNE for vizualization**

```python
#first split quality
red_wine_x = red_wine.iloc[:,:11]
red_wine_y = red_wine.iloc[:,-1]
```

```python
red_wine_y
```

```python
#first standardizing data
from sklearn.preprocessing import StandardScaler
std_red_wine = StandardScaler().fit_transform(red_wine_x)
```

```python
#using t-SNE alogorithm for vizualization
from sklearn.manifold import TSNE
model = TSNE(n_components=2, perplexity=50, random_state=0)
tsne_data = model.fit_transform(std_red_wine)
tsne_data = np.vstack((tsne_data.T, red_wine_y)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "quality"))
sns.set_style("darkgrid")
sns.FacetGrid(tsne_df, hue='quality', height=10).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.show()
```

```python
#changing perplexity and iterations for t-SNE algorithm
from sklearn.manifold import TSNE
model = TSNE(n_components=2, perplexity=30, n_iter=5000, random_state=0)
tsne_data = model.fit_transform(std_red_wine)
tsne_data = np.vstack((tsne_data.T, red_wine_y)).T
tsne_df = pd.DataFrame(data=tsne_data, columns=("Dim_1", "Dim_2", "quality"))
sns.FacetGrid(tsne_df, hue='quality', height=10).map(plt.scatter, 'Dim_1', 'Dim_2').add_legend()
plt.show()
```

From the above t-SNE vizualization, we see that the quality scores 5 and 6 are spread all over and quality scores 7 and 8 are cornered and distinguishable.

# **Red Wine Quality Prediction Using k-Nearest Neighbors**

<h2>1. Importing libraries</h2>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn import model_selection
```

<h2>2. Loading dataset into pandas DataFrame</h2>

```python
#first break the dataset into training, cross validation and testing dataset

red_wine = pd.read_csv('/kaggle/input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
```

```python
red_wine
```

<h2>3. Separating the target from the features</h2>

```python
#separating 'quality column'
red_wine_x = red_wine.iloc[:,:-1]
red_wine_y = red_wine.iloc[:,-1]
```

```python
red_wine_x
```

<h2>4. Column Standardization</h2>

```python
#column standardization
from sklearn.preprocessing import StandardScaler
red_wine_x = StandardScaler().fit_transform(red_wine_x)
```

<h2>5. Simplifying classes for kNN</h2>

```python
#simplifying classes for kNN 
red_wine_y_copy = red_wine_y.copy()
red_wine_y = red_wine_y.replace([3,4],0)
red_wine_y = red_wine_y.replace([5,6],1)
red_wine_y = red_wine_y.replace([7,8],2)  

red_wine_y.value_counts()
```

<h2>6. Converting into a numpy array</h2

```python
X = np.array(red_wine_x)
Y = np.array(red_wine_y)
```

<h2>7. Splitting data into train, cross validation and test datasets and fitting the model</h2>

```python
# split the data set into train and test
X_1, X_test, Y_1, Y_test = model_selection.train_test_split(X,Y,test_size = 0.3, random_state=0)

#split X_1, Y_1 into training and cross validation datasets
x_train, x_cv, y_train, y_cv = model_selection.train_test_split(X_1, Y_1, test_size = 0.3)

#7-NN
knn = KNeighborsClassifier(n_neighbors=7)

#fitting data on training datasets
knn.fit(x_train, y_train)

#predict on x cross validation
predict_cv = knn.predict(x_cv)

#accuracy on y cross validation
accuracy_cv = accuracy_score(y_cv, predict_cv, normalize=True) * float(100)
print(f"Cross Validation accuracy for k=7 is {accuracy_cv}")
```

```python
#now testing on X_test
knn = KNeighborsClassifier(n_neighbors=7)

knn.fit(x_train, y_train)
#testing on X_test
predict_test = knn.predict(X_test)

#accuracy on Y_test
accuracy_test = accuracy_score(Y_test, predict_test, normalize=True) * float(100)
print(f"Test dataset accuracy for k=7 is {accuracy_test}")
```

```python
#predicting on train set to verify overfitting
predict_train = knn.predict(x_train)

accuracy_train = accuracy_score(y_train, predict_train, normalize=True) * float(100)
print(f"Train dataset accuracy for k=7 is {accuracy_train}")
```

Therefore, no overfitting
