```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv("../input/shanghai-lane-house-rentals-2021/for_tableau_2.csv")
```

```python
data.describe()
```

```python
data.head()
```

```python
plt.figure(figsize=(25,10))
sns.countplot(x='district' , data=data)
```

```python
data['rent'].value_counts()
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(data['sqmeters'],data['rent'], color='red')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig=plt.figure(figsize=(10,10))
ax=Axes3D(fig)
ax.scatter(data['rent'],data['bedrooms'],data['sqmeters'])
plt.xlabel("Rent")
plt.ylabel("Bedrooms")
plt.show()
```
