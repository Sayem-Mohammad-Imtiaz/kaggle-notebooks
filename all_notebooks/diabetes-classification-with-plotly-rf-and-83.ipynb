# Step 3: Modelling

**Let's try RandomForest and Logistic Regression.**

```python
y = healthset[['Outcome']]
X = healthset.drop(columns=['Outcome']) #[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       #'BMI', 'DiabetesPedigreeFunction', 'Age']]

# T-T-S
X_train,X_test,y_train,y_test = train_test_split(X, y, train_size=0.8,random_state=42)

# Random Forest Model
forest_model = RandomForestClassifier(n_estimators=5)
forest_model.fit(X_train,y_train.values.ravel())
```

```python
print("\n Random Forest Model Score: ",forest_model.score(X_test,y_test))

y_predicted = forest_model.predict(X_test)

#print("\n Y_test : ",y_test)
#print("\n Y_Predicted : ",y_predicted)

print("\n Random Forest Model R-Squared : ", r2_score(y_test, y_predicted) * 100)
```

```python
from sklearn.metrics import confusion_matrix

#plotting a confusion matrix for prediction
conmat = confusion_matrix(y_test, y_predicted)

print("\n Confusion matrix: \n", conmat)

conmat_fig = plex.imshow(conmat)
conmat_fig.show()
```
