```python
df= pd.read_csv('/kaggle/input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
```

# Drop some duplicates from the dataset

```python
df= df.drop_duplicates(subset='Name')
```

# 1. How many non-fictional or fictional books?

```python
popular_genre= df['Genre'].value_counts().reset_index()
popular_genre= popular_genre.rename(columns={"index":"Genre", "Genre":"Count"})
popular_genre
```

```python
fig = px.pie(popular_genre, names='Genre', values='Count')
fig.update_layout(title_text='Which Genre is more than the other?')
fig.show()
```

There are 310 non-fictional and 240 fictional books

# 2. Which author release more books?

```python
best_author= df[['Name', 'Author']].groupby('Author')['Name'].count().sort_values(ascending=False).head(10).reset_index()
best_author
```

```python
fig = px.bar(best_author, x='Author', y='Name')
fig.update_layout(title_text='Which author released more books?')
fig.show()
```

# 3. Most released books in 1 year by single author

```python
best_year= df[['Name', 'Year', 'Author']].groupby(['Year','Author'])['Name'].count().sort_values(ascending=False).reset_index()

best_year = best_year.drop_duplicates(subset='Year')
```

```python
fig = px.bar(best_year, x='Year', y='Name', color= 'Author')
fig.update_layout(title_text='Most productive author for each year (2009-2019)')
fig.show()
```

# 4. Which author have more reviews? 

```python
most_review_author= df[['Reviews', 'Author']].groupby('Author')['Reviews'].sum().sort_values(ascending=False).reset_index(drop=False).head(10)
most_review_author
```

```python
fig = px.bar(most_review_author, x='Author', y='Reviews')
fig.update_layout(title_text='Which author received more reviews?')
fig.show()
```

# 5. Which author has the best rating among them?

```python
best_rating_author= df[['Author', 'User Rating']].groupby('Author')['User Rating'].mean().sort_values(ascending= False).reset_index(drop=False).head(10)
best_rating_author
```

# 6. Which Author make cheaper or more expensive books?

### Cheaper

```python
cheapest_author= df[['Author', 'Price']].groupby('Author')['Price'].mean().sort_values().reset_index(drop= False).head(10)
cheapest_author
```

```python
fig = px.bar(cheapest_author, x='Author', y='Price')
fig.update_layout(title_text='Which author made cheaper books')
fig.show()
```

### More expensive

```python
expensive_author= df[['Author', 'Price']].groupby('Author')['Price'].mean().sort_values(ascending=False).reset_index(drop=False).head(10)
expensive_author
```

```python
fig = px.bar(expensive_author, x='Author', y='Price')
fig.update_layout(title_text='Which author made more expensive books')
fig.show()
```

# 7. Which Genre got more reviews?

```python
popular_genre= df[['Genre','Reviews']].groupby('Genre')['Reviews'].sum().reset_index(drop=False)

popular_genre
```

```python
fig= px.pie(popular_genre, names= 'Genre', values='Reviews')
fig.update_layout(title_text='Which Genre received more reviews?')
fig.show()
```

# 8. Which genre have better rating?

```python
genre_rating= df[['Genre', 'User Rating']].groupby('Genre')['User Rating'].mean().reset_index(drop=False)
genre_rating
```

```python
fig= px.pie(genre_rating, names= 'Genre', values='User Rating')
fig.update_layout(title_text='Which Genre better rating?')
fig.show()
```

# 9. Price average for each year and genre

```python
price_avg_year = df[['Price', 'Year', 'Genre']].groupby(['Genre', 'Year'])['Price'].mean().reset_index(drop=False)
price_avg_year
```

```python
fig= px.bar(price_avg_year, x='Year', y='Price',color='Genre', barmode='group')
fig.update_layout(title_text='Average books price for each year')
fig.show()
```

# 10. Does higher rating means higher price?

```python
rating_vs_price = df[['User Rating', 'Price']].groupby('User Rating')['Price'].mean().reset_index(drop=False)
rating_vs_price
```

```python
fig=px.line(rating_vs_price, x= 'User Rating', y='Price')
fig.update_layout(title_text='Does higher rating means higher price?')
fig.show()
```

# 11. Which books got the most reviews

```python
most_reviewed_books= df[['Name', 'Reviews']].groupby('Name')['Reviews'].sum().sort_values(ascending=False).head(10).reset_index(drop=False)
most_reviewed_books
```

```python
fig=px.bar(most_reviewed_books, x='Name', y='Reviews')
fig.update_layout(title_text='Which books got more reviews?')
fig.show()
```
