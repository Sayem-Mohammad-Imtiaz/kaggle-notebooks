## Data cleaning

For columns Embarked has 2 rows missing data. We will drop the rows of missing data.
For columns Cabin has too much missing data. We will drop columns Cabin.
And columns Age has 20 percent missing data. We will fill with mean age of all passengers. 

```python
mean_age = round(taitanic['Age'].mean())
mean_age
```

drop columns Cabin

```python
taitanic = taitanic.drop(columns = ['Cabin'])
```

Filling nan in Age columns by mean_age

```python
age = taitanic['Age'].replace(math.nan,mean_age)
taitanic['age'] = age
taitanic = taitanic.drop(columns = ['Age'])
```

```python
taitanic.head(100)
```

Converting Categorical Features

```python
taitanic.info()
```

```python
sex = pd.get_dummies(taitanic['Sex'],drop_first=True)
embark = pd.get_dummies(taitanic['Embarked'],drop_first=True)
```

```python
taitanic.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
```

```python
taitanic = pd.concat([taitanic,sex,embark],axis=1)
```

```python
taitanic.head(100)
```

## Train Test Split

```python
X = taitanic.drop('Survived', axis = 1)
Y = taitanic['Survived']
```

## Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.30, random_state=101)
```

## Training and Predicting

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

## Evaluation

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
