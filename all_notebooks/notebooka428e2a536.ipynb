```python
from glob import glob
import os
from tqdm import tqdm
import pandas as pd
```

```python
files = glob('../input/notebook5558080c89/runs/detect/exp/labels/*.txt')
```

```python
len(files)
```

```python
dictresult = {}
```

```python
for file in tqdm(files):
    file1 = open(file, 'r')
    Lines = file1.readlines()
    bbox = ""
    for i in range(len(Lines)):
        a = list(map(float,Lines[i].split()))
        x,y,xm,ym = a[2],a[3],a[4]+a[2],a[1]+a[3]
        x = round(x*1024)
        y = round(y*1024)
        xm = round(xm*1024)
        ym = round(ym*1024)
        bbox += " ".join(list(map(str,[x,y,xm,ym])))
        bbox += ";"

    dictresult[os.path.split(file)[1][:-4]] = bbox
```

```python
df = pd.Series(dictresult).to_frame('PredString')
```

```python
df = df.reset_index()
df.columns = ['image_name','PredString']
```

```python
sample_sub = pd.read_csv('../input/aicrowdwheatdata/submission.csv')
sample_sub
```

```python
sub = sample_sub.merge(df, on="image_name", how="left")
```

```python
sub
```

```python
sub['PredString'] = sub.PredString_y.fillna('no_box')
```

```python
sub[['image_name','domain','PredString']].to_csv('submissionfold1newyolov5.csv',index=False)
```
