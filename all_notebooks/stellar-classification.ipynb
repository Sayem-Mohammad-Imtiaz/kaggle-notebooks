```python
df = pd.read_csv("/kaggle/input/star-categorization-giants-and-dwarfs/Star3642_balanced.csv")
```

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
ord_enc.fit_transform(df)
```

```python
X = df.drop('SpType', axis=1)
```

```python
df
```

```python
X_ = X.drop('TargetClass', axis = 1)
```

```python
y_ = X['TargetClass']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_,y_)
```

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42)
```

```python
sgd_clf.fit(X_train, y_train)
```

```python
y_pred = sgd_clf.predict(X_test)
```

```python
n_correct = sum(y_pred == y_test)
```

```python
sgd_accu = n_correct/len(y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

ran_for = RandomForestClassifier()
ada_clf = AdaBoostClassifier()

ran_for.fit(X_train,y_train)
ada_clf.fit(X_train,y_train)
```

```python
y_pred_ran = ran_for.predict(X_test)
y_pred_ada = ada_clf.predict(X_test)
```

```python
ada_accu = sum(y_pred_ada==y_test)/len(y_pred_ada)
ran_accu = sum(y_pred_ran==y_test)/len(y_pred_ran)
```

```python
print(sgd_accu,ada_accu,ran_accu)
```
