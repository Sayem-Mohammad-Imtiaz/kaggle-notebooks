Data Preprocessing 

sklearn.model_selection 
out of which used ones are 
* train_test_split --> to split train and test data
* KFold --> for K-fold cross validator
* cross_val_predict and cross_val_score 
* GridSearchCV and RandomizedSearchCV for hyperparameter tuning 



```python
X=boston.drop('MEDV',axis=1)
Y=boston['MEDV']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
X_test.shape
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.coef_ , lr.intercept_
```

```python
lr.score(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
y_train_pred = lr.predict(X_train)
```

```python
y_test_predict=lr.predict(X_test)
```

```python
residual_test=((y_test_predict-y_test)**2).sum()
```

```python
np.sqrt(residual_test/X.shape[0])
```
