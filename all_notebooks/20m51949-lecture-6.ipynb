We look at Corona cases for a specific country.

```python
import numpy as np # linear algebra
import matplotlib.pyplot as plt  # visualization
import pandas as pd  # data processing
np.set_printoptions(threshold=np.inf)

selected_country='Denmark'
df = pd.read_csv('../input/novel-corona-virus-2019-dataset/covid_19_data.csv',header=0) #header=0 to set the first row as the header
#print(np.unique(df['Country/Region']cor.values))
df = df[df['Country/Region']==selected_country]
df = df.groupby('ObservationDate').sum()
print(df)
```

Note: This df is actually a cumulative information.

# To calculate the daily increases

```python
df['daily_confirmed']=df['Confirmed'].diff() #diff() is to substract the row before
df['daily_deaths']=df['Deaths'].diff()
df['daily_recovered']=df['Recovered'].diff()
df['daily_confirmed'].plot()
df['daily_deaths'].plot()
df['daily_recovered'].plot()
plt.show()
```

```python
print(df)
```

# Interactive chart

To do this, we'll need to load three of the following modules.

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_confirmed_object = go.Scatter(x=df.index,y=df['daily_confirmed'].values,name='Daily confirmed')
daily_deaths_object = go.Scatter(x=df.index,y=df['daily_deaths'].values,name='Daily deaths')
daily_recovered_object = go.Scatter(x=df.index,y=df['daily_recovered'].values,name='Daily recovered')

layout_object = go.Layout(title=selected_country + ' daily cases 20M51949',xaxis=dict(title='Date'),yaxis=dict(title='Number of people'))
fig = go.Figure(data=[daily_confirmed_object,daily_deaths_object,daily_recovered_object],layout=layout_object)
iplot(fig)
fig.write_html(selected_country + ' daily_cases_20M51949.html')
```

```python
from plotly.offline import iplot
import plotly.graph_objs as go

confirmed_object = go.Scatter(x=df.index,y=df['Confirmed'].values,name='Cumulative confirmed')
deaths_object = go.Scatter(x=df.index,y=df['Deaths'].values,name='Cumulative deaths')
recovered_object = go.Scatter(x=df.index,y=df['Recovered'].values,name='Cumulative recovered')

layout_object = go.Layout(title=selected_country + ' cumulative cases 20M51949',xaxis=dict(title='Date'),yaxis=dict(title='Number of people'))
fig = go.Figure(data=[confirmed_object,deaths_object,recovered_object],layout=layout_object)
iplot(fig)
fig.write_html(selected_country + ' daily_cases_20M51949.html')
```

# Discussion of the Denmark responses to COVID-19 pandemic

> #  ***Short summary: The Danish Governmentâ€™s response to COVID-19 pandemic has been successful.***

# **Impression**

Since March, I have heard the news about the large economic supports of the danish government due to the impacts of the COVID-19 pandemic. I felt very impressed on how reactive their government is at the time. Two months later, I have also seen news presenting the different approaches of each Scandinavian country such as Denmark, Norway, and Sweden. Briefly, Denmark and Norway acted quickly in implementation of various treatment and preventive measures against virus spreading, but Sweden implemented much less tried to keep the society functions as usual as possible. Today, the results prove it. The number of new cases in Denmark and Norway greatly reduce, while the number of new cases in Sweden keeps growing exponentially at this moment.

# **Suppression Strategy**

The suppression strategy aims for social distancing and lockdown. The target is to prevent the spreading of viruses inside the country.

* **Febraury 27**: The first case in Denmark was confirmed.
* **March 11**: The first death was confirmed.  
* **March 13**: The emergency law was enacted. It is mainly to restrict the access to public facilities and transport. It also enhances the powers of public health authorities. 
* **March 14**: Country borders are closed. It is allowed to only residents and foreigners who are working in Denmark.
* **March 16**: Lockdown is enforced.

As we can see from the timeline, the Danish government has swift responses in protecting their residences and preventing virus spreading expansion. Several measures are quickly introduced and implemented. Technically, Denmark is also one of the first European countries to lock down the country border. The quarantine measures are also strongly enforced. All people return from high-risk areas, all infected people and people who have been in contact with infected people are required to quarantine themselves. 

# **Enhacing health care system**

It needs to thank the health care system of Denmark (and several European countries) which is mainly the public sector with the concept of universal access to the health care system. For example, the Danish government funds 85% of costs. It is different from the system in the US and UK which are mostly the private sector. 

However, Denmark also encountered the same risk as other countries that the capacity of the health care system cannot support a large amount of infected in a short period of time. For example, the number of intensive care beds in Denmark is around 6.7 per 100000 inhabitants, while the average of European countries is 11.5. Therefore, "Flattening the curve" must be achieved otherwise the system will collapse, and situation will be getting worse. Several guidelines for self-care during COVID-19 have been issued. For example, washing hands, avoiding physical contact, and social distancing. Besides, the diagnosis and treatment of each health care system have been improved.


# **Mitigating economic damages**

Several companies surely encounter the difficulty during the COVID-19 pandemic. Danish government introduced various financial aid and economic measures to protect the economy and support employees. They spent up to approximately EUR 54 billion. The example of the important measures is the temporary wage compensation (state 75%, company 25%), and the fixed expenses compensation and the postponement of VAT and payroll taxes of each company.
