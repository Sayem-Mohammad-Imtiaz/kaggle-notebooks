```python
df = pd.read_csv('/kaggle/input/solar-power-generation-data/Plant_1_Generation_Data.csv')
df.head()
```

```python
df.info()
```

```python
df1 = df[df.columns[3:]]
df1
```

```python
df1.corr()
```

```python
# Hubungan Korelasi setiap Kolom
import seaborn as sns
ax = sns.heatmap(df1.corr(),annot = True,linewidths=.75,cmap='viridis_r')
ax.set_title('Hubungan Nilai Korelasi Tiap Kolom')
```

```python
# Grafik nilai hubungan tiap kolom
import matplotlib.pyplot as plt
fig,ax = plt.subplots(3,2,figsize=(20,20))
ax[0][0].scatter(x = df['DC_POWER'],y =df['AC_POWER'])
ax[0][0].set_title('Hubungan antara DC POWER dan AC Power',fontsize = 15)
ax[0][0].set_xlabel('DC POWER',fontsize = 13)
ax[0][0].set_ylabel('AC POWER',fontsize = 13)
ax[0][1].scatter(x = df['DAILY_YIELD'],y =df['TOTAL_YIELD'], color = 'r')
ax[0][1].set_title('Hubungan antara DAILY YIELD dan TOTAL YIELD',fontsize = 15)
ax[0][1].set_xlabel('DAILY YIELD',fontsize = 13)
ax[0][1].set_ylabel('TOTAL YIELD',fontsize = 13)
ax[1][0].scatter(x = df['DAILY_YIELD'],y =df['AC_POWER'], color = 'g')
ax[1][0].set_title('Hubungan antara DAILY YIELD dan AC POWER',fontsize = 15)
ax[1][0].set_xlabel('DAILY YIELD',fontsize = 13)
ax[1][0].set_ylabel('AC POWER',fontsize = 13)
ax[1][1].scatter(x = df['TOTAL_YIELD'],y =df['AC_POWER'], color = 'y')
ax[1][1].set_title('Hubungan antara TOTAL YIELD dan AC POWER',fontsize = 15)
ax[1][1].set_xlabel('TOTAL YIELD',fontsize = 13)
ax[1][1].set_ylabel('AC POWER',fontsize = 13)
ax[2][0].scatter(x = df['DAILY_YIELD'],y =df['DC_POWER'], color = 'c')
ax[2][0].set_title('Hubungan antara DAILY YIELD dan DC POWER',fontsize = 15)
ax[2][0].set_xlabel('DAILY YIELD',fontsize = 13)
ax[2][0].set_ylabel('DC POWER',fontsize = 13)
ax[2][1].scatter(x = df['TOTAL_YIELD'],y =df['DC_POWER'], color = 'm')
ax[2][1].set_title('Hubungan antara TOTAL YIELD dan DC POWER',fontsize = 15)
ax[2][1].set_xlabel('TOTAL YIELD',fontsize = 13)
ax[2][1].set_ylabel('DC POWER',fontsize = 13)
plt.show()
```

>KESIMPULAN
Hubungan korelasi antara DC POWER dan AC POWER mendekati 1 dan cenderung positif (high possitive correlation). Artinya semakin meningkatnya nilai DC POWER, nilai dari AC POWER juga cenderung meningkat. Sedangkan korelasi antara kolom lainnya mendekati 0 atau tidak berkorelasi
