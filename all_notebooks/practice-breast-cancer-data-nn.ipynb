```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load in 

import numpy as np # linear algebra
from matplotlib import pyplot as plt
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix,classification_report
# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory
from sklearn.neural_network import MLPClassifier
from subprocess import check_output
print(check_output(["ls", "../input"]).decode("utf8"))
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
# Any results you write to the current directory are saved as output.
```

```python
class BreastCancer:
    def load_data(self):
        self.cancer=load_breast_cancer()
        self.X=self.cancer['data']
        self.Y=self.cancer.target
        self.train,self.test,self.train_y,self.test_y=train_test_split(self.X,self.Y)
        self.g_fit=[]
        pass
    def StandardScaling(self):
        scaler=StandardScaler()
        scaler.fit(self.X)
        scaler.transform(self.train)
        scaler.transform(self.test)
        pass
    def sklearnMLP(self,layers_list):
        for layer_req in layers_list:
            classifier=MLPClassifier(hidden_layer_sizes=layer_req,random_state=7)
            classifier.fit(self.train,self.train_y)
            self.predictions=classifier.predict(self.test)
            print('Evaluation for the model with',layer_req)
            self.evaluate_model()
            
            pass
    def evaluate_model(self):
        self.confusion_mat=confusion_matrix(self.test_y,self.predictions)
        self.class_report=classification_report(self.test_y,self.predictions)
        #print('Confusion mat:')
        #print(self.confusion_mat)
        #print('Classification report:')
        #print(self.class_report)
        #print('Correlation:')
        #print(np.power(np.corrcoef(self.predictions,self.test_y),2)[0,1])
        #return np.power(np.corrcoef(self.predictions,self.test_y),2)[0,1]
        self.g_fit.append(np.power(np.corrcoef(self.predictions,self.test_y),2)[0,1])
        pass 
        
        
```

```python
bc1=BreastCancer()
bc1.load_data()
bc1.StandardScaling()
```

```python
for i in range(1,500):
    bc1.sklearnMLP([(i,)])
```
