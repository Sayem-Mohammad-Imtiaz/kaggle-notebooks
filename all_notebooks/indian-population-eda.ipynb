# We can apply stats model in it. 

```python
import statsmodels.api as sm
x = df['Population'].tolist()
y = df['Year'].tolist()
x = sm.add_constant(x)  
result = sm.OLS(y, x).fit()
print(result.summary())
```

```python
import statsmodels.api as sm
x = df['GrowthRate'].tolist()
y = df['Year'].tolist()
x = sm.add_constant(x)  
result = sm.OLS(y, x).fit()
print(result.summary())
```

# We can see Adjusted R square and R Square from it. 

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
X = df[['Year']]
Y = df[['Population']]
model = lr.fit(X,Y)
Pred_X = pd.DataFrame(list(range(2021,2090)),columns=['Year'])
Pred_X['Prediction'] = model.predict(Pred_X[['Year']])
Pred_X
```
