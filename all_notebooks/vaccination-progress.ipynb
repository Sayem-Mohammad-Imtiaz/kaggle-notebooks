**What country vaccinated more people?**

```python
vaccinations = df.groupby(by="country").sum()
vaccinations.head()
```

```python
vaccinations.sort_values(by="total_vaccinations", ascending=False, inplace=True)
vaccinations.head()
```

**Here we obsereved United states has vaccinated more people**

**What country is vaccinated a larger percent from its population?**

```python
people_vaccinated = df[['country', 'total_vaccinations_per_hundred']]
people_vaccinated.head()
people_vaccinated.max
```

```python
people_vacc_grouped = people_vaccinated.groupby("country").mean()
people_vacc_grouped.head()
```

```python
people_vacc_grouped.sort_values(by="total_vaccinations_per_hundred", ascending=False, inplace=True)
people_vacc_grouped.head()
```
