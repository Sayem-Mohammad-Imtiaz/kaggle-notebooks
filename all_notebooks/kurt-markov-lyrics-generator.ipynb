```python
from collections import defaultdict, Counter
import random
import sys
import os
import pandas as pd
import csv

# number of words the markov chain makes predictions based on
words = 8

lyrics = ''
df = pd.read_csv('../input/nirvana-lyrics/raw_data.csv')
for index, row in df.iterrows():
    song = row.to_dict()
    lyrics = lyrics +'\n'+song['lyrics']

kurt = defaultdict(Counter)

for i in range(len(lyrics) - words):
    state = lyrics[i:i + words]
    next = lyrics[i + words]
    kurt[state][next] += 1
state = random.choice(list(kurt))
out = list(state)


for i in range(30):
    for i in range(30):
        out.extend(random.choices(list(kurt[state]), kurt[state].values()))
        state = state[1:] + out[-1]
print('\n'.join(''.join(out).split(sep=',')))

```
