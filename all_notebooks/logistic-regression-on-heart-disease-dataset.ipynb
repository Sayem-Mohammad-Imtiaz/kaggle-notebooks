```python
# FROM THE CODES BELOW YOU CAN ANALYSE THAT THE ACCURACY RATE IS HIGH IF RELATABLE COLUMNS ARE TAKEN INTO CONSIDERATION. 

# LOGISTIC REGRESSION ON 'heart_disease' DATASET TO PREDICT RESULT
# WE WILL TAKE FACTORS FROM 'age TO 'thal'

import pandas as pd
from sklearn import model_selection
import statsmodels.api as sm
from sklearn.metrics import accuracy_score


df=pd.read_csv('../input/heart-disease-patients-details/heart_disease.csv')
Y=df['result']
X=df.loc[:,'age':'thal']


Xtrain,Xtest,Ytrain,Ytest=model_selection.train_test_split(X,Y,random_state=4,test_size=0.33)

logit=sm.Logit(Ytrain,Xtrain)
result=logit.fit()

abc=result.predict(Xtest)

print(abc)
print(Ytest)
```

```python
# LOGISTIC REGRESSION ON heart_disease DATASET TO PREDICT RESULT
# WE WILL TAKE FACTORS FROM 'age' TO 'number_of_major_vessels'
# ACCURACY RATE IS LITTLE LOW AS COMPARED TO ABOVE AS 'thal' COLUMN IS NOT TAKEN IN X

import pandas as pd
from sklearn import model_selection
import statsmodels.api as sm
from sklearn.metrics import accuracy_score


df=pd.read_csv('../input/heart-disease-patients-details/heart_disease.csv')
Y=df['result']
X=df.loc[:,'age':'number_of_major_vessels']


Xtrain,Xtest,Ytrain,Ytest=model_selection.train_test_split(X,Y,random_state=4,test_size=0.33)

logit=sm.Logit(Ytrain,Xtrain)
result=logit.fit()

abc=result.predict(Xtest)

print(abc)
print(Ytest)
```
