```python
import pandas as pd
import numpy as np
import seaborn as sns
data = pd.read_csv('../input/yelp-reviews-dataset/yelp.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data['text length'] = data['text'].apply(len)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
sns.boxplot(x='stars',y='text length',data = data)
```

```python
sns.countplot(x='stars',data=data,palette = 'rainbow')
```

```python
stars = data.groupby('stars').mean()
stars
```

```python
sns.heatmap(stars.corr(),cmap='coolwarm',annot=True)
```

```python

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()

```

```python
new_data = data[(data['stars']==1) | (data['stars']==5)]
new_data
X = new_data['text']
y = new_data['stars']
X = cv.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33)
```

```python
#training a model
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
nb.fit(X_train,y_train)
```

```python
predictions = nb.predict(X_test)
```

```python
#creating a confusion matrix
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(y_test,predictions))

print(classification_report(y_test,predictions))
```

```python
from sklearn.feature_extraction.text import  TfidfTransformer
```

```python
from sklearn.pipeline import Pipeline
```

```python
#Creating pipeline
pipeline = Pipeline([
    ('bow', CountVectorizer()),  # strings to token integer counts
    ('tfidf', TfidfTransformer()),  # integer counts to weighted TF-IDF scores
    ('classifier', MultinomialNB()),  # train on TF-IDF vectors w/ Naive Bayes classifier
])
```

```python
X = new_data['text']
y = new_data['stars']
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state=101)
```

```python
pipeline.fit(X_train,y_train)
```

```python
predictions = pipeline.predict(X_test)
```

```python
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```
