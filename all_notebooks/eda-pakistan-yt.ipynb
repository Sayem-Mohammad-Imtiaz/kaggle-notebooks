# What Pakistan is looking at

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../input/pakistans-top-250-youtube-channels/Pakistan Top 250 YouTube Channels.csv')
df
```

```python
df.drop('Top #', axis=1, inplace=True)
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df['Twitter '].str.split(",",expand = True) 
df.head(5)
```

```python

df = df.rename(columns={0: 'Channel', 1: 'Category', 2: 'Subscribers'})
df
```

```python
def subs_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    
df['Subscribers'] = df['Subscribers'].apply(subs_to_float)
df.head()
```

```python
df.isnull().sum()
```

```python
df['Category'].unique()
```

```python
plt.figure(figsize=(17, 7))
plt.xticks(rotation=60)
graph = sns.barplot(x='Category', y='Subscribers', data=df)
```

**Pakistan is Definately watching Entertainment (Dramas) more than any other**

```python
plt.figure(figsize=(17, 7))
plt.xticks(rotation=60)
graph = sns.barplot(x='Channel', y='Subscribers', data=df.query('Subscribers > 5000000'))
```

**Channels with more than 5 Million Subscribers**

```python
plt.figure(figsize=(10, 10))
channels = df.query('Category == "Comedy"')
subs = list(channels['Subscribers'])
names = list(channels['Channel'])
plt.pie(subs, labels=names, autopct='%1.1f%%', colors=['#c4ff75', '#baff9e', '#ff9c59'])
plt.show()
```
