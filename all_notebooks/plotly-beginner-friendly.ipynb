### **INTRO**



- In this study, we are going to make Exploratory Data Analysis (EDA) with the Life Expectancy dataset.
- Study aims to be beginner friendly and give as much as possible explanation for each step on the way.
- Study's dataset has 3253 instances along with country name, life expectancy, and year.
- Dataset includes life expectancy records between 1800 and 2016. 217 Years of data, quite impressive.

- First thing first, lets' import the related libraries for further analysis

```python
import pandas as pd
import numpy as np

import plotly 
import plotly.express as px
import plotly.graph_objs as go
import plotly.offline as py
from plotly.offline import iplot
from plotly.subplots import make_subplots
import plotly.figure_factory as ff
```

```python
df = pd.read_csv('../input/countries-life-expectancy/Life expectancy.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['Year'].unique()
```

- We have a data between the years of 1800 and 2016.

```python
df['Entity'].unique()
```

- Also we have 15 different countries in the dataset.

```python
df.info()
```

- Everything seems quite OK.

- Let's move on to the analysis part.

### Analysis Part

#### **Life Expectancy**

```python
df['Life expectancy'].describe()
```

- Quite expected.
- We have 216 years of data on life expectancy from different continents and it is very normal to see skewed distribution.
- Based on the descriptive info, we can expect highly right skewed data.
- Let's see it

```python
fig = px.histogram(df, x= 'Life expectancy', title='Life Expectancy', marginal="box", hover_data = df[['Entity','Year']])
fig.show()
```

- As we expected, even though median score is almost 42 years, distribution density increases at around 50 years.
- And we have a quite a lot of records of life expectancy bigger than median value of life expectancy.

### Life Expectancy By Year

```python
px.scatter(df, x="Year", y="Life expectancy", color='Entity')
```

- We have a lot of useful information in this scatter plot.

- Let's see life expectancy by year in animated version of bar chart.  

```python
fig = px.bar(df, x="Life expectancy",y="Entity", color="Entity",
  animation_frame="Year", animation_group="Entity", range_x=[0,85])
fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 30
fig.layout.updatemenus[0].buttons[0].args[1]['transition']['duration'] = 5
fig.show()
```

- As seen in the both scatterplot and animated bar chart,
   - Life expectancy is at it's lowest points at the beginning of the 1800 
   - And during the World War 1, especially at 1918
   - and during the World War 2
   - Significant life expectancy scores can be observed among the countries till 1950s.
   - After 1950s, life expectancy scores start to increase significantly throughout the dataset.
   

- Let's see all of these observations closely in the different continents.

### **Europe**

```python
df_Europe = df[df['Entity'].isin(['France', 'Germany','Italy','Spain','Switzerland', 'United Kingdom'])].copy()
df_Europe["Life expectancy"].describe()
```

#### Life Expectancy By Year in Europe

- First start with the scatterplot

```python
px.scatter(df_Europe, x="Year", y="Life expectancy", color='Entity')
```

- In the data analysis, using visualization is very important and most of the times, if it is used effectively, it would be an excellent help for a data analyst.

- In this scatter plot, during the World War 1 and World War 2, sudden decreases in the life expectancy scores can be observed clearly. 

- Let's use animated version of the bar chart and see these changes in the action.

```python
fig = px.bar(df_Europe, x="Life expectancy", y="Entity",color="Entity",
  animation_frame="Year", animation_group="Entity", range_x=[0,85])
fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 30
fig.layout.updatemenus[0].buttons[0].args[1]['transition']['duration'] = 5
fig.show()
```

- Ok it is much more clear now that, in Europe, during the World War 1 and World War 2, 'life expectancy' was at the lowest points.
