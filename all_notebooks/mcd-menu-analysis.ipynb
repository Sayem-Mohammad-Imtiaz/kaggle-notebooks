```python
from scipy.stats import zscore
```

```python
mcdnum1 = mcdnum.apply(zscore)
```

```python
cov_mat = np.cov(mcdnum1.T)
eig_vals,eig_vecs = np.linalg.eig(cov_mat)
tot =sum(eig_vals)
var = [(i/tot)*100 for i in sorted(eig_vals,reverse=True)]
cumsumvar = np.cumsum(var)
```

```python
cumsumvar
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=7)
mcdred = pca.fit_transform(mcdnum1)
```

```python
mcdred1 = pd.DataFrame(mcdred)
```

```python
mcdred1
```

```python
wsc=[]
```

```python
from sklearn.cluster import KMeans
for i in range(1,16):
    kmeans= KMeans(n_clusters=i)
    kmeans.fit(mcdred1)
    wsc.append(kmeans.inertia_)
    
```

```python
plt.plot(range(1,16),wsc)
```

Let us consider 4 clusters.

```python
from sklearn.metrics import silhouette_samples,silhouette_score
```

```python
kmeans4= KMeans(n_clusters=4)
kmeans4.fit(mcdred1)
kmeans4.labels_
silhouette_score(mcdred1,kmeans4.labels_)
```

```python
mcdnum["cluster"]= kmeans4.labels_
```

```python
fig, axes = plt.subplots(nrows=22,ncols=1)
fig.set_size_inches(15,120)
j=0
for i in mcdnum.columns:
    
    
    a = sns.scatterplot(y=mcdnum[i] ,x=mcdnum["cluster"], ax=axes[j])
    
    a.set_title(i,fontsize=12)
    j= j + 1
plt.show()
```
