### Importing Libraries

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", category=RuntimeWarning)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
%matplotlib inline
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix
```

### Importing Dataset

```python
data = pd.read_csv('../input/breast-cancer-wisconsin-data/data.csv')
```

```python
data.drop(data.columns[[32]], axis = 1, inplace = True) 
data.head()
```

```python
data.drop(['id'], axis=1, inplace=True)
```

### Data Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['diagnosis'] = le.fit_transform(data['diagnosis'])
le.classes_
```

### Splitting data into train and test sets.

```python
from sklearn.model_selection import StratifiedShuffleSplit


feature_cols = [x for x in data.columns if x != 'diagnosis']


# Split the data into two parts with 1500 points in the test data
# This creates a generator
strat_shuff_split = StratifiedShuffleSplit(n_splits=1, test_size=450, random_state=42)

# Get the index values from the generator
train_idx, test_idx = next(strat_shuff_split.split(data[feature_cols], data['diagnosis']))

# Create the data sets
X_train = data.loc[train_idx, feature_cols]
y_train = data.loc[train_idx, 'diagnosis']

X_test = data.loc[test_idx, feature_cols]
y_test = data.loc[test_idx, 'diagnosis']
```

```python
y_train.value_counts(normalize=True)
```

```python
y_test.value_counts(normalize=True)
```

### Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,roc_auc_score

DecisionTreeClassifier_model = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
DecisionTreeClassifier_model.fit(X_train, y_train)
DecisionTreeClassifier_predict=DecisionTreeClassifier_model.predict(X_test)
DecisionTreeClassifier_accuracy = accuracy_score(y_test, DecisionTreeClassifier_predict)
```

```python
rp=classification_report(y_test, DecisionTreeClassifier_predict)
print("Classification Report:")
print()
print(rp)
disp = plot_confusion_matrix(DecisionTreeClassifier_model, X_test, y_test,
                                                    cmap= plt.cm.Blues)
```
