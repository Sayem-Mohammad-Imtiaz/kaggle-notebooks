```python
df['Popularity'].isnull().sum()
```

```python
new_order = sorted(df['Chord'].dropna().unique())
new_order
```

```python
# drop 11 rows with incomplete information (Nan values or empty space)

df2 = df[df['Chord'].map(lambda x:x in new_order)]
df2.info()
```

```python
sns.heatmap(df2.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
df2.columns
```

```python
sns.countplot(x='Popularity',data=df2)
```

```python
import math
df2['new_Popularity'] = df2['Popularity'].apply(lambda x: int(math.floor(x/10)))
```

```python
ax = sns.countplot(x='Chord',data=df2,order=new_order,hue='new_Popularity')
ax.legend(bbox_to_anchor=(1.0,1.0))
```

```python
df2['new_Popularity'].value_counts().plot(kind='pie',autopct='%.2f',figsize=(20,6))
```

```python
sns.jointplot(x='Popularity',y='Number of Times Charted',data=df2)
```

```python
ax= sns.heatmap(df2.corr(),annot=True,cmap='YlGnBu')
```

```python
from datetime import datetime
df2['Release_Date'] = pd.to_datetime(df2['Release Date'])
df2['Release_Date']
```

```python
df2['Year'] = df2['Release_Date'].dt.year
df2['Year']
```

```python
df2['Month'] = df2['Release_Date'].dt.month
df2['Month']
```

```python
df2.drop('Release Date',inplace=True,axis=1)
```
