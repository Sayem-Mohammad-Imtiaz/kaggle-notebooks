```python
import pandas as pd 
import numpy as np 
```

```python
data = pd.read_csv('../input/mushroom-classification/mushrooms.csv')
data.head()
```

```python
print('data shape -->',data.shape)
data.info()
```

```python
cols = data.columns
print(cols)
```

```python
for i in cols:
  print('unique values for colmun :',i,data[i].unique())
  print('='*40)
```

```python
data.drop('veil-type',axis = 1,inplace= True)
cols = data.columns
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for i in cols:
  le.fit(data[i])
  data[i] = le.transform(data[i])
```

```python
print(data.info())
```

```python
X = data.drop(['class'],axis = 1)
y = data['class']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.33,shuffle = True)
```

```python
from sklearn.linear_model import LogisticRegression
LogisticRegressionModel = LogisticRegression(penalty='l2',solver='sag',C=1.0,random_state=33,max_iter = 2000)
LogisticRegressionModel.fit(X_train, y_train)

#Calculating Details
print('LogisticRegressionModel Train Score is : ' , LogisticRegressionModel.score(X_train, y_train))
print('LogisticRegressionModel Test Score is : ' , LogisticRegressionModel.score(X_test, y_test))
```

```python
from sklearn.svm import SVC
SVCmodel= SVC(kernel= 'rbf', max_iter=10000,C=1.0,gamma='auto')
SVCmodel.fit(X_train,y_train)
print(SVCmodel.score(X_test,y_test))
print(SVCmodel.score(X_train,y_train))
```
