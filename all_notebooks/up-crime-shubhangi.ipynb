```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../input/up-crime-data-year-2019/MR data - Compiled Data Set.csv')
```

```python
df.head(10)
```

```python
df=df.drop(['Title'],axis=1)
```

```python
df=df.drop(['Text'],axis=1)
```

```python
df.head(10)
```

```python
df=df.fillna(0)
```

```python
df.shape
```

```python
columns=df.columns
```

```python
df.columns=[i for i in range(len(df.columns))]
```

```python
df.head(20)
```

```python
dic={0:0,'Property Disputes':1,'Family Dispute':2,'Petty Quarrels':3,'Money Disputes':4, 'Personal Vendetta':5, 'Love Affairs':6, 'Casteism':7,'Unknown reasons':8}
```

```python
df[1]=df[1].apply(lambda x : dic[x])
```

```python
df.describe()
```

```python
df[11]=0
df[12]=0
```

```python
for i in range(0,len(df)):
    if ',' in str(df.iloc[i,8]):
        cl=df.iloc[i,8].split(',')
        df.iloc[i,8]=int(cl[0])
        df.iloc[i,11]=int(cl[1])
        if len(cl)>2:
            df.iloc[i,12]=int(cl[2])
    elif int(df.iloc[i,8])>100:
        df.iloc[i,11]=int(df.iloc[i,8])%100
        df.iloc[i,8]=int(df.iloc[i,8])//100
```

```python
df.head(20)
```

```python
df.describe()
```

```python
g=df.groupby(0)
df1=g.get_group('Ghaziabad')
df2=g.get_group('Kanpur')
df3=g.get_group('Lucknow')
```

```python
group2=df.groupby(1)
Pr=group2.get_group(1)
Fam=group2.get_group(2)
Pet=group2.get_group(3)
Mon=group2.get_group(4)
Per=group2.get_group(5)
La=group2.get_group(6)
Cast=group2.get_group(7)
Unk=group2.get_group(8)
```

```python
len(Pr)
```

```python
Reasons=["Property/Land Disputes","Family Dispute","Petty Quarrels","Money Disputes","Personal Vendetta","Love Affairs","Casteism","Unknown/other"]
Number_of_Crimes=[len(Pr),len(Fam),len(Pet),len(Mon),len(Per),len(La),len(Cast),len(Unk)]
```

```python
print(Number_of_Crimes)
```

```python
plt.bar(Reasons,Number_of_Crimes,color='blue',width=0.5)
plt.title("Number of crimes on the basis of Reasons ")
plt.xlabel("Reasons")
plt.ylabel("Number of Crimes")
plt.xticks(rotation = 90)
plt.show()
```

The most number of crimes occur due to unknown reasons and the second most prominent reason is due to Love Affairs .

```python
Child=[df1[2].sum(),df2[2].sum(),df3[2].sum()]
Male=[df1[3].sum(),df2[3].sum(),df3[3].sum()]
Female=[df1[4].sum(),df2[4].sum(),df3[4].sum()]
city=["Ghaziabad","Kanpur","Lucknow"]
```

```python

  
X_axis = [0.0,2.0,4.0]
X_1=[0.4,2.4,4.4]
X_2=[0.8,2.8,4.8] 
plt.bar(X_axis,Child,width=0.4,color='red',label='Child Victims')
plt.bar(X_1,Male,width=0.4,color='green',label='Male Victims')
plt.bar(X_2,Female,width=0.4,color='blue',label='Female Victims')
  
plt.xticks(X_1,city)
plt.xlabel("City")
plt.ylabel("Number of Victims")
plt.title("Number of Victims in each city")
plt.legend()
plt.show()
    

```

Lucknow has the most number of crimes with adult victims(both male and female) but has the least number of 
child crimes among the three cities .
