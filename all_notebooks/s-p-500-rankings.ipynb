```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("/kaggle/input/sp-500-companies/constituents-financials.csv")
df.head()
```

```python
df.info()
```

```python
#To check the data 
df.isnull()
```

```python
#To find subsectors in 'Sector'column
df['Sector'].unique()
```

```python
#To visualize the size of each subsector
plt.figure(figsize=(8,8))
sns.countplot(y=df['Sector'])
plt.title("Size of Each Sector")
plt.show()
```

```python
#To rename column Name to Company
df = df.rename(columns={'Name':'Company'})
df.head()
```

```python
#To find Top 10 companies of each category
def getTop10Companies(category,returnData=False):
    plt.figure(figsize=(15,15))
    top10 = df[['Company',category]].sort_values(by=category,ascending=False).head(10)
    graph=sns.barplot(data=top10,y='Company',x=category)
    plt.title(f'Top 10 companies in {category}')
    plt.show()
    
    if returnData:
        return top10
```

```python
for col in df.columns[2:]:
    getTop10Companies(col)
    print('\n')
```
