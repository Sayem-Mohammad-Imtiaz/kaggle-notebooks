```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
```

**<font color =Orange size = 5>Load data</font>**

```python
df=pd.read_csv('/kaggle/input/paralmpic-2020/Paralympic_Medal_Count.csv')
```

**<font color =Orange size = 5>Data analysis</font>**

```python
df1 = df.sort_values(by = ['total medals'],ascending = False).reset_index().head(10)

tr1 = go.Bar(x = df1['Team/Npc'],
             y = df1['Gold'],
             name = 'Gold',
             marker = dict(color = 'Gold'))

tr2 = go.Bar(x = df1['Team/Npc'],
             y = df1['silver'],
             name = 'Silver',
             marker = dict(color = 'Silver'))

tr3 = go.Bar(x = df1['Team/Npc'],
             y = df1['bronze'],
             name = 'Bronze',
             marker = dict(color = 'Peru'))

data = [tr1,tr2,tr3]

layout = dict(title ='Top 10 Countries with the most Paralympic Medals',
              title_x = 0.5,
              title_font = dict(size = 16, color = 'Maroon'),
              barmode = 'relative',
              xaxis = dict(title = 'Country'),
              yaxis = dict(title = 'Count')
                  
                  )
fig = go.Figure(data = data, layout = layout)
fig.show()
```

**<font color =Coral size = 4>Animated Map</font>**

```python
df1 = df.sort_values(by = ['total medals'],ascending = False).reset_index().head(10)

fig = px.choropleth(df1, 
                    locations = 'Team/Npc',
                    locationmode = 'country names',
                    color = 'total medals',
                    hover_name = 'Team/Npc',
                    color_continuous_scale = 'plasma',
                    animation_frame = 'Team/Npc',
                    hover_data = ['rank by total'],
                    labels = {'total medals':'Total Medals','Team/Npc':'Country','rank by total':'Rank'}
                   
                   )

fig.update_layout(title = 'World Countries with Tokyo Paralympic Medals',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'MidnightBlue'),
                  geo= dict(showframe = False,
                            showcoastlines = False,
                            projection_type = 'equirectangular'
                           ))

fig.show()
```

```python
df1 = df.groupby("Team/Npc")['Gold'].sum().sort_values(ascending = False).reset_index().head(10)

fig = px.bar(df1,x = 'Team/Npc', y = 'Gold',color ='Gold',color_continuous_scale = 'blues',
             )
fig.update_layout(title = 'Top 10 Countries with the most Gold Medals',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'DarkSlateBlue'),
                  xaxis = dict(title = 'Country')
                 
                 )

fig.show()
```

**<font color =Coral size = 4>Animated Map</font>**

```python
df1 = df.sort_values(by = ['Gold'],ascending = False).reset_index()

fig = px.choropleth(df1,
                    locations = 'Team/Npc',
                    locationmode = 'country names',
                    color = 'Gold',
                    color_continuous_scale = 'blues',
                    animation_frame = 'Team/Npc',
                    labels = {'total medals':'Total Medals','Team/Npc':'Country'}
                    
                   )

fig.update_layout(title = 'World Countries with Gold Medals',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'DarkSlateBlue'),
                  geo = dict(showframe = False,
                             showcoastlines = False,
                             projection_type = 'equirectangular'
                            
                            )
                 
                 
                 )

fig.show()
```
