### Example Data Visualization

Here we extract 3 "Neighbourhoods" within the Greater Toronto Area and plot some simple charts to visualize the housing price growth over the years. We will generate 4 hours:

* Trend of Housing prices in Markham
* Trend of Housing prices in Richmond Hill
* Trend of Housing prices in Aurora
* Trend of Apartment Prices across Markham, Richmond Hill, and Aurora

```python
#Extract "Markham" data
Markham=HousingData[HousingData['Location']=='Markham'].copy()
Markham.reset_index(drop=True, inplace=True)

RichmondHill=HousingData[HousingData['Location']=='Richmond Hill'].copy()
RichmondHill.reset_index(drop=True, inplace=True)

Aurora=HousingData[HousingData['Location']=='Aurora'].copy()
Aurora.reset_index(drop=True, inplace=True)
```

```python
fig, axs = plt.subplots(nrows=4,ncols=1,figsize=(20,20))
axs[0].plot(Markham['Date'], Markham['SFDetachBenchmark'], label="SingleFamily Detach")
axs[0].plot(Markham['Date'], Markham['SFAttachBenchmark'], label="SingleFamily Attach")
axs[0].plot(Markham['Date'], Markham['THouseBenchmark'], label="Townhouse")
axs[0].plot(Markham['Date'], Markham['ApartBenchmark'], label="Apartment")
axs[0].plot(Markham['Date'], Markham['CompBenchmark'], label="Composite")
axs[0].legend()
axs[0].set_title('Markham')

axs[1].plot(RichmondHill['Date'], RichmondHill['SFDetachBenchmark'], label="SingleFamily Detach")
axs[1].plot(RichmondHill['Date'], RichmondHill['SFAttachBenchmark'], label="SingleFamily Attach")
axs[1].plot(RichmondHill['Date'], RichmondHill['THouseBenchmark'], label="Townhouse")
axs[1].plot(RichmondHill['Date'], RichmondHill['ApartBenchmark'], label="Apartment")
axs[1].plot(RichmondHill['Date'], RichmondHill['CompBenchmark'], label="Composite")
axs[1].legend()
axs[1].set_title('Richmond Hill')

axs[2].plot(Aurora['Date'], Aurora['SFDetachBenchmark'], label="SingleFamily Detach")
axs[2].plot(Aurora['Date'], Aurora['SFAttachBenchmark'], label="SingleFamily Attach")
axs[2].plot(Aurora['Date'], Aurora['THouseBenchmark'], label="Townhouse")
axs[2].plot(Aurora['Date'], Aurora['ApartBenchmark'], label="Apartment")
axs[2].plot(Aurora['Date'], Aurora['CompBenchmark'], label="Composite")
axs[2].legend()
axs[2].set_title('Aurora')

axs[3].plot(Markham['Date'], Markham['ApartBenchmark'], label="Markham")
axs[3].plot(RichmondHill['Date'], RichmondHill['ApartBenchmark'], label="Richmond Hill")
axs[3].plot(Aurora['Date'], Aurora['ApartBenchmark'], label="Aurora")
axs[3].legend()
axs[3].set_title('Apartment Benchmark Price Comparison')
```
