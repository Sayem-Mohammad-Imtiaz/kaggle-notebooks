```python
chess = pd.read_csv("../input/worlds-chess-grandmasters-data/WorldChessGrandMaster.csv",parse_dates=['Born'])
chess.head()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=chess.Federation.value_counts()[:10].index,y=chess.Federation.value_counts()[:10].values)
for i,j in enumerate(chess.Federation.value_counts().sort_values(ascending=False)[:10].values):
    plt.text(i-0.2,j+1,str(j),fontdict={'fontsize':10,'color':'black'})
plt.title("Top-10 Countries Of GrandMasters")
plt.xlabel("Countries")
plt.ylabel("No Of GrandMasters")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=[str(i) for i in chess['Title Year'].value_counts().sort_values(ascending=False)[:10].index],y=chess['Title Year'].value_counts().sort_values(ascending=False)[:10].values)
for i,j in enumerate(chess['Title Year'].value_counts().sort_values(ascending=False)[:10].values):
    plt.text(i,j+1,str(j),fontdict={'fontsize':10,'color':'black'})
plt.title("Top-10 Years For Chess GrandMasters")
plt.xlabel("Year")
plt.ylabel("No Of GrandMasters")
plt.show()
```

```python
plt.figure(figsize=(12,8))
ax=plt.subplot()
for col in chess.Federation.value_counts().sort_values(ascending=False)[:5].index:
    sns.lineplot(x=chess[chess.Federation == col]['Title Year'].value_counts().index,y=chess[chess.Federation == col]['Title Year'].value_counts().values,ax=ax,marker="o",legend=[col])
plt.legend(chess.Federation.value_counts().sort_values(ascending=False)[:5].index)
plt.title("Top-5 Countries Year Wise GrandMaster Title Achievements")
plt.xlabel("Year")
plt.ylabel("No of GrandMasters")
plt.show()
```

```python
chess.Sex.value_counts().plot.pie(radius=2,autopct="%.2f",center=(20,0),textprops={'fontsize':15},pctdistance=0.8)
plt.title("Gender Dominace In Chess GrandMasters",y=1.38,fontdict={'fontsize':20})
plt.ylabel(None)
plt.legend(['Male','Female'])
plt.show()
```

<p style="font-size:40px;"> Youngest GrandMasters</p>

**Note:** Considering only year of born and the year in which they got GrandMaster title. Result may be different if we consider actual date i.e Month and year also but at the same time we don't have actual title date.

```python
# Separating Born years and subtracting it from Title year to get the period to achieve this GrandMaster title

chess['Born_Year'] = chess.Born.dt.year
chess['Age_To_Achive_Title'] = chess['Title Year'].astype('int') - chess['Born_Year']
chess.head()
```

```python
chess[chess.Age_To_Achive_Title == chess.Age_To_Achive_Title.min()]
```

<h1 style="font-size:40px;">Most Amazing Part Of Whole Visualization i.e Bar Chart Race</h1>

<h2>Q. What is Bar Chart Race?</h2>

<p>Ans. The animated bar chart race helps us to visualize the change in trends over time, these type of charts are very popular on social media as they provide a holistic data story/insight in a concise and easy to understand chart.</p>

<h2>Q. What we will need to create bar chart race?</h2>

1. We will need bar_chart_race libray to create BarChartRace. So we need to install it and import it into our working enviornment.

2. To create a a chart race, data must be in 'wide' format where each row represents a single time period and each column represents a distinct category. Optionally, the index can label the time period.So we need to create a new wide range dataframe to create bar chart race.

**Below `chart_race` dataframe is a wide range format**

```python
chart_race = pd.DataFrame(index=np.sort(chess['Title Year'].unique()))
chart_race[chess.Federation.unique()] = 0
chart_race.head()
```

```python
#inserting values in respective columns wrt to time frame.

for time in np.sort(chess['Title Year'].unique()):
    chart_race.loc[time,chess[chess['Title Year'] == time]['Federation'].value_counts().index] = chess[chess['Title Year'] == time]['Federation'].value_counts().values
chart_race.head()
```
