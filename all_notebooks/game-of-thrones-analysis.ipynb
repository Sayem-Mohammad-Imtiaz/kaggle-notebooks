# Game of Thrones

Game of Thrones is an American fantasy drama television series created by David Benioff and D. B. Weiss for HBO. It is an adaptation of A Song of Ice and Fire, a series of fantasy novels by George R. R. Martin, the first of which is A Game of Thrones. The show was shot in the United Kingdom, Canada, Croatia, Iceland, Malta, Morocco, and Spain. It premiered on HBO in the United States on April 17, 2011, and concluded on May 19, 2019, with 73 episodes broadcast over eight seasons.

In this notebook, let's see how the character deaths are distributed

```python
import pandas as pd
import numpy as np
import json

import plotly.express as px
import plotly.graph_objects as go

import warnings
warnings.filterwarnings('ignore')
```

```python
deaths = pd.read_csv("../input/d/prashant111/game-of-thrones/character_deaths.csv")
deaths.head()
```

```python
deaths.info()
```

```python
# Removing 'House' from Allegiances
deaths['allegiances'] = deaths['allegiances'].str.replace('House ', '')
```

```python
deaths_by_houses = deaths.groupby('allegiances').size()
deaths_by_houses.drop(['None', "Night's Watch", 'Wildling'], inplace=True)
deaths_by_houses = deaths_by_houses.reset_index(name='Deaths')
deaths_by_houses = deaths_by_houses.rename(columns = {'allegiances': 'House'})
deaths_by_houses
```

```python
fig = px.bar(deaths_by_houses, x='House', y='Deaths',
             title="Total Deaths of Name Characters by House",
             color='House',
             color_discrete_sequence=px.colors.sequential.Tealgrn,
             template='plotly_white')

fig.layout.xaxis.fixedrange = True
fig.layout.yaxis.fixedrange = True
fig.show()
```

```python
deaths_by_gender = deaths.groupby('gender').size().reset_index(name='Deaths')
deaths_by_gender.gender[deaths_by_gender.gender == False] = 'Female'
deaths_by_gender.gender[deaths_by_gender.gender == True] = 'Male'
deaths_by_gender = deaths_by_gender.rename(columns = {'gender': 'Gender'})
deaths_by_gender
```

```python
fig = px.bar(deaths_by_gender, x='Gender', y='Deaths',
             color='Gender',
             color_discrete_sequence=['#1e3d59', '#ff6e40'],
             title="Total Deaths by Gender",
             template='plotly_white')

fig.layout.xaxis.fixedrange = True
fig.layout.yaxis.fixedrange = True
fig.show()
```

```python
deaths_by_nobility = deaths.groupby('nobility').size().reset_index(name='Deaths')
fig = px.pie(deaths_by_nobility, values='Deaths', names='nobility',
             title='Total Deaths by Nobility',
             color_discrete_sequence=['#1e3d59', '#ff6e40'],
             template='plotly_white')

fig.layout.xaxis.fixedrange = True
fig.layout.yaxis.fixedrange = True
fig.show()
```

The death distribution is almost on equal terms between Nobility and common people. The common folks have slightly more deaths than nobility.

```python
season_1 = open('../input/game-of-thrones-srt/season1.json', 'r')
season_2 = open('../input/game-of-thrones-srt/season2.json', 'r')
season_3 = open('../input/game-of-thrones-srt/season3.json', 'r')
season_4 = open('../input/game-of-thrones-srt/season4.json', 'r')
season_5 = open('../input/game-of-thrones-srt/season5.json', 'r')
season_6 = open('../input/game-of-thrones-srt/season6.json', 'r')
season_7 = open('../input/game-of-thrones-srt/season7.json', 'r')

seasons = [season_1, season_2, season_3, season_4, season_5, season_6, season_7]
```

```python
length_season_dic = {}
for season in range(len(seasons)):
    i = season + 1
    name = 'Season ' + str(i)
    length_season_dic[name] = len(seasons[season].read())
length_season_dic
```

```python
length_season = {'Seasons': length_season_dic.keys(),
                 'Lines of Sentences': length_season_dic.values()}
length_season = pd.DataFrame(length_season)


fig = go.Figure(data=go.Scatter(x=length_season["Seasons"], y=length_season["Lines of Sentences"]))

fig.update_layout(title_text='Sentences Spoken Per Season',
                 template='plotly_white')

fig.update_xaxes(title_text='Seasons')
fig.update_yaxes(title_text='Sentences')

fig.layout.xaxis.fixedrange = True
fig.layout.yaxis.fixedrange = True
fig.show()
```

As we can see from the above line graph, the sentences spoken in the show drastically decrease as the season progresses.
