### 1) Veri Setinin Yüklenmesi ve İlk Bakış

```python
maindata = pd.read_csv("../input/covid19-data-set-of-turkey/tr-covid19.csv", index_col = 0)
maindatac = maindata.copy()
```

```python
# Veri setinin sütunlarındaki isimlendirmede boşluk bulunduğundan boşluk yerine '_' gelecek şekilde replace ettik.

maindatac.columns = maindatac.columns.str.replace(' ','_')
```

```python
# Sütun isimlerini küçük harf yaptık

maindatac.columns= maindatac.columns.str.lower()
```

### 3) Boş ve Aykırı Değerlerin Bulunması

#### Eksik Değerlerin Bulunması ve Doldurulması

```python
maindatac.isnull().sum()
```

```python
# Total_number_of_test sütuunun doldurulması
maindatac["total_number_of_test"].fillna( maindatac.groupby("month")["total_number_of_test"].transform("mean"), inplace = True)

# Total_number_of_deaths sütuunun doldurulması
maindatac["total_number_of_deaths"].fillna( maindatac.groupby("month")["total_number_of_deaths"].transform("mean"), inplace = True)

# Number_of_critically_ill_patients sütuunun doldurulması
maindatac["number_of_critically_ill_patients"].fillna( maindatac.groupby("month")["number_of_critically_ill_patients"].transform("mean"), inplace = True)

# Total_Number_of_recoveries sütuunun doldurulmasıa
maindatac["total_number_of_recoveries"].fillna( maindatac.groupby("month")["total_number_of_recoveries"].transform("mean"), inplace = True)

# number_of_cases sütuunun doldurulması
maindatac["number_of_cases"].fillna( maindatac.groupby("month")["number_of_cases"].transform("mean"), inplace = True)

# number_of_sick sütuunun doldurulmasıa
maindatac["number_of_sick"].fillna( maindatac.groupby("month")["number_of_sick"].transform("mean"), inplace = True)

# number_of_test sütuunun doldurulması
maindatac["number_of_test"].fillna( maindatac.groupby("month")["number_of_test"].transform("mean"), inplace = True)

# number_of_deaths sütuunun doldurulmasıa
maindatac["number_of_deaths"].fillna( maindatac.groupby("month")["number_of_deaths"].transform("mean"), inplace = True)

# Number_of_recoveries sütuunun doldurulması
maindatac["number_of_recoveries"].fillna( maindatac.groupby("month")["number_of_recoveries"].transform("mean"), inplace = True)
```

```python
maindatac["total_number_of_test"].astype(int)
maindatac["total_number_of_deaths"].astype(int)
maindatac["number_of_critically_ill_patients"].astype(int)
maindatac["total_number_of_recoveries"].astype(int)
maindatac["number_of_cases"].astype(int)
maindatac["number_of_sick"].astype(int)
maindatac["number_of_test"].astype(int)
maindatac["number_of_deaths"].astype(int)
maindatac["number_of_recoveries"].astype(int)

```

### 5) Analiz

```python
X = maindatac.drop(["number_of_cases", "number_of_sick","number_of_recoveries","number_of_test","number_of_deaths","date","month","year"], axis=1)
y = maindatac["total_number_of_deaths"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)

print ("X_train", X_train.shape)
print ("X_test", X_test.shape)
print ("y_train", y_train.shape)
print ("y_test", y_test.shape)
```

```python
reg = LinearRegression()
model = reg.fit(X_train, y_train)
print ("Sabit Katsayı: ", model.intercept_)
print("Değişkenlerin Katsayıları: ", model.coef_)
```

```python
lm = sm.OLS(y_train, X_train)
model = lm.fit()
model.summary()
```
