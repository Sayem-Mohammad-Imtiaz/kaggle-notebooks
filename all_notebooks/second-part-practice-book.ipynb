```python
def proportion_of_education():
#-------------------------------------------------------------------
    import pandas as pd
    df = pd.read_csv('../input/second-dataset/NISPUF17.csv')
    T = len(df['EDUC1'])
    a, b, c, d = 0, 0, 0, 0
#-------------------------------------------------------------------
    for num, value in df['EDUC1'].iteritems():
        if (value == 1):
            a += 1
        if (value == 2):
            b += 1
        if (value == 3):
            c += 1
        if (value == 4):
            d += 1
#-------------------------------------------------------------------
    p = {"less than high school" :a/T,
        "high school" :b/T,
        "more than high school but not college": c/T,
        "college" :d/T}
#-------------------------------------------------------------------
    return p
    raise NotImplementedError()
```

```python
proportion_of_education()
```

```python
def average_influenza_doses():
    # YOUR CODE HERE
    import pandas as pd
    df = pd.read_csv('../input/second-dataset/NISPUF17.csv')
    #df['CBF_01'].unique()
    df = df[['CBF_01','P_NUMFLU']]
    df = df[df['P_NUMFLU'].notna()]
    #df
    T_1 = len(df[df['CBF_01'] == 1])
    T_2 = len(df[df['CBF_01'] == 2])
    m = df[df['CBF_01'] == 1]['P_NUMFLU'].sum()
    nm = df[df['CBF_01'] == 2]['P_NUMFLU'].sum()
    #Total_1, Total_2
    return (m/T_1, nm/T_2)
    raise NotImplementedError()
```
