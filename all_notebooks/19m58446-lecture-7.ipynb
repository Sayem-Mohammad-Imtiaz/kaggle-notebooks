# Create a global map of COVID 19 vaccination (ratio (in %) between population fully vaccinated and total population)

### The animated map below shows the percentage of the countries' population that have been fully vaccinated. I am interested in this particular data because I think the percantage to total population could be an objective indicator to compare each country's performance in vaccine rollouts.

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
#print(df.columns)
#print(df)

fig = px.choropleth(df,locations='country',locationmode='country names',color='people_fully_vaccinated_per_hundred',hover_name='country',animation_frame='date',color_continuous_scale='greens',range_color=(0.0,40.0))
fig.update_layout(title_text="Percentage of Countries' Total Population that are Fully Vaccinated",title_x=0.1)
fig.show()

plotly.offline.plot(fig,filename='vaccination.html')
```
