```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

df=pd.read_csv('../input/backtest/bat.csv')
df
```

```python
#Probility density for Run-up before reaching SL point
step = 0.01
ot=0
do=df[df['Profit %']<0]['Run-up %'].max()
bin_range = np.arange(ot,do+step, step)
count, division = np.histogram(df['Run-up %'],bins =bin_range)
prob=100*count/df.shape[0]
prob[0:3].sum()
df1=pd.DataFrame(prob,division[1:],columns=['Freq of '+str(step)+' bin'] )

sumprob=[]
for i in range(len(division[1:])):
  sumprob.append(100-prob[0:i].sum())
df1['Density']=sumprob

df1.plot(figsize=(40,10),grid=True)
plt.xticks(np.arange(0,df1.index.max(),0.1),fontsize=18)
plt.yticks(np.arange(0,100,5),fontsize=20)
plt.show()
```
