```python
#6) Ayrıca bu safhada kategorik değişkenlerin sıralanabilir(ordered) veya sıralanamaz(nominal) 
#olma durumlarına göre uygun sayısallaştırma işlemleri yapılmalıdır.
df.head()
```

```python
label_encoder = preprocessing.LabelEncoder()
```

```python
df['Bluetooth'] = label_encoder.fit_transform(df['Bluetooth'])
df['CiftHat'] = label_encoder.fit_transform(df['CiftHat'])
df['4G'] = label_encoder.fit_transform(df['4G'])
df['3G'] = label_encoder.fit_transform(df['3G'])
df['Dokunmatik'] = label_encoder.fit_transform(df['Dokunmatik'])
df['WiFi'] = label_encoder.fit_transform(df['WiFi'])
```

```python
df['Renk'] = pd.Categorical(df['Renk'])
dfDummies = pd.get_dummies(df['Renk'], prefix = 'Renk')
dfDummies
```

```python
df = pd.concat([df, dfDummies], axis=1)
df.head()
```

```python
df.drop(["Bluetooth","CiftHat","4G","3G" ,"WiFi" ,"Renk" ,"Dokunmatik"], axis = 1, inplace = True)
```

```python
df.head()
```

```python
#7) İkinci safhada makine öğrenmesi modellemeleri yapabilmek için hedef değişkenimiz olan “FiyatAraligi” değişkeni
#ve bu değişkeni tahmin etmemize yardımcı olacak diğer değişkenler bağımlı ve bağımsız değişkenler olmak üzere ayrılmalıdır
x = df.drop("FiyatAraligi", axis = 1)#bağımlı
y = df["FiyatAraligi"] #bağımsız
y
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,
                                                    test_size=0.15, 
                                                    random_state=42)
```

```python
#9) Eğitim kümesi ile GaussianNB, DecisionTree ve KNN modelleri eğitilmeli ve her bir model için karmaşıklık matrisi çizdirilmeli,
#doğruluk skoru elde edilerek modeller arasındaki başarı durumu karşılaştırma yapılmalıdır.
nb = GaussianNB()
nb_model = nb.fit(x_train, y_train)
```

```python
#10) . Modellerin başarısını değerlendirme konusunda F1_Score, Precision, Recall değerleri “classification_report” aracılığıyla yazdırılmalıdır.
F1_score = f1_score(y_test, y_pred, average = 'weighted')#önceki ikinin ağırlık ortalaması
F1_score
```

```python
print(classification_report(y_test, y_pred))
```
