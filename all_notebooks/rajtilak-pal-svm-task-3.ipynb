```python
df=pd.read_csv('/kaggle/input/breast-cancer-wisconsin-data/data.csv')
```

```python
df.head()
```

```python
X=df.iloc[:,2:-1]
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

## Applying Plain SVM

```python
svm=SVC()
```

```python
svm.fit(X_train,y_train)
```

```python
y_pred_train=svm.predict(X_train)
y_pred_test=svm.predict(X_test)
```

### Accuracy Scores 


```python
print("Training Accuracy: ",accuracy_score(y_train,y_pred_train))
print("Testing Accuracy: ",accuracy_score(y_test,y_pred_test))
```

## Applying GridSearchCV

```python
svm=SVC()
```

```python
param_grid=[]
```

```python
param_grid.append({'kernel': ['linear'], 'C': [0.01,0.1,1,10,100,200,500,1000]})
param_grid.append({'kernel': ['poly'], 'C': [0.01,0.1,1,10,100,200,500,1000], 'degree': range(2,11)})
param_grid.append({'kernel': ['rbf'], 'C': [0.01,0.1,1,10,100,200,500,1000]})
param_grid
```

```python
gd=GridSearchCV(svm,param_grid,cv=10,n_jobs=-1,scoring='accuracy')
```

```python
gd.fit(X_train,y_train)
```

### Best Params

```python
gd.best_params_
```

### Best Estimator

```python
gd.best_estimator_
```

### Best Score

```python
gd.best_score_
```

## Taking the best estimator to predict values

```python
svm=gd.best_estimator_
```

```python
y_pred_train=svm.predict(X_train)
y_pred_test=svm.predict(X_test)
```

## Final Training and Testing Accuracies 

```python
print("Training Accuracy: ",accuracy_score(y_train,y_pred_train))
print("Testing Accuracy: ",accuracy_score(y_test,y_pred_test))
```
