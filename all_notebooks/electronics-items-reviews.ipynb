#### Replacing nan values for *title* and *review_body* with empty spaces ""

```python
data['title'] = data['title'].fillna("")
data['Review body'] = data['Review body'].fillna("")
```

```python
data = data.sample(frac=1).reset_index(drop=True)
```

## Creating WordCloud for positive reviews and negative reviews

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image

import matplotlib.pyplot as plt
% matplotlib inline
```

```python
plt.figure(figsize=(20, 20))
body_pos = " ".join(review for review in data[data['sentiment'] == 'positive']['Review body'])
stopwords = set(STOPWORDS)

wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(body_pos)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(20, 20))
body_neg = " ".join(review for review in data[data['sentiment'] == 'negative']['Review body'])
stopwords = set(STOPWORDS)

wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(body_pos)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
