Now letz start making our model

```python
data.columns
```

```python
X = data.drop(["output"],axis =1)
y = data.output
```

```python
cat_cols = []
num_cols = []

for col in X.columns:
    if(X[col].nunique() <= 5):
        cat_cols.append(col)
    else:
        num_cols.append(col)
print(f"categorical:{cat_cols}\n"
      f"Numerical:{num_cols}")
```

```python
X = pd.get_dummies(X, columns=cat_cols, drop_first=True)
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
data = RandomForestClassifier(random_state = 3)
data.fit(X_train,y_train)

predict_train_data = data.predict(X_test)
acc1 = accuracy_score(y_test,predict_train_data)
accuracy1 = acc1*100
print(str(accuracy1)+"%")
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)

pred = model.predict(X_test)
acc2 = accuracy_score(y_test,pred)
accuracy2 = acc2*100
print(str(accuracy2)+"%")
```
