```python
import pandas as pd
from fastai.text.all import *
```

# Data Preprocessing:

The CSVs given are filled with data that I wouldn't want to use right now so I will make two news CSVs: train and test. 

Apart from that, we'll need to creat a seperate CSv for our labels.

```python
train = pd.read_csv("../input/uci-drug/UCIdrug_train.csv")
test = pd.read_csv("../input/uci-drug/UCIdrug_test.csv")
```

Let's take out what we need to build a classifier:

```python
train = pd.concat([train['review'], train['rating']], 
                  axis = 1)

test = pd.concat([test['review'], test['rating']], 
                 axis = 1)

train.to_csv("./train.csv")
test.to_csv("./test.csv")
```

Lets grab all the unique values:

```python
array = train.rating.unique()
labels = pd.DataFrame(array)
labels.to_csv("./labels.csv")
labels
```

Creating a Dataloader:

```python
dls = TextDataLoaders.from_csv(path = "./", csv_fname= 'train.csv',
                               test_col='review', label_col= 'rating',
                               valid_col=None)
# Lets take a look at it:
dls.show_batch(max_n=3)
```

# Training the Model:
Now this is the fun part.

Defining the learner:

```python
learn = text_classifier_learner(dls, AWD_LSTM, metrics=accuracy)
```

```python
learn.fine_tune(3, 3e-4)
```

```python
learn.fine_tune(3, 1e-2)
```

# Predicting

```python
learn.predict(['you absolutely suck, your stuff sucks'])
```
