```python
%%writefile tfpactice01.html
<!DOCTYPE HTML>
<html>
  <head>
    <meta charset='utf-8'>
    <meta name='viewport' content='width=device-width'>
    <title>üèô TF 01 SMC</title>
    <script src='https://sagecell.sagemath.org/static/embedded_sagecell.js'></script>
    <script>$(function(){
      sagecell.makeSagecell(
        {inputLocation:'div.linked',linked:true,evalButtonText:'run linked cells'}); });
    </script>
  </head>
  <style>
    @import 'https://fonts.googleapis.com/css?family=Akronim|Aladin';
    h2 {color:#348ABD; font-family:Akronim; font-size:120%; 
        text-shadow:4px 4px 4px #aaa;}
    p {color:slategray; font-family:Aladin; font-size:120%; 
       text-shadow:4px 4px 4px #aaa;}
    .sagecell .CodeMirror-scroll {min-height:3em; max-height:70em;}
    body {margin:5px 5px 5px 15px;}
  </style>
  <body>
    <h2>‚úíÔ∏è &nbsp; Code Modules</h2>
<div class='linked'><script type='text/x-sage'>
!python3 -m pip install --upgrade tensorflow_hub \
--user --quiet --no-warn-script-location
path='/home/sc_work/.sage/local/lib/python3.8/site-packages'
import sys; sys.path.append(path)
import warnings; warnings.filterwarnings('ignore')
import os,urllib,PIL.Image,tensorflow_hub as hub
import numpy as np,tensorflow as tf,pylab as pl
from IPython.display import display,HTML
file_path='https://olgabelitskaya.gitlab.io/images/'
tfhub_path='https://tfhub.dev/google/magenta/'+\
           'arbitrary-image-stylization-v1-256/2'
os.environ['TFHUB_MODEL_LOAD_FORMAT']='COMPRESSED'
</script></div><br/>
     <h2>‚úíÔ∏è &nbsp; Loading Images</h2>      
<div class='linked'><script type='text/x-sage'>
def get_file(file,file_path=file_path):
    input_file=urllib.request.urlopen(file_path+file)
    output_file=open(file,'wb'); 
    output_file.write(input_file.read())
    output_file.close(); input_file.close()
def load_img(path_to_img):
    img=tf.io.read_file(path_to_img)
    img=tf.image.decode_image(img,channels=3)
    img=tf.image.convert_image_dtype(img,tf.float32)
    shape=tf.cast(tf.shape(img)[:-int(1)],tf.float32)
    scale=int(512)/max(shape)
    new_shape=tf.cast(shape*scale,tf.int32)
    img=tf.image.resize(img,new_shape)
    return img[tf.newaxis,:]
def tensor2img(tensor):
    tensor=np.array(tensor*int(255),dtype=np.uint8)
    if np.ndim(tensor)>3:
        assert tensor.shape[0]==1; tensor=tensor[0]
    return PIL.Image.fromarray(tensor)
def display_images(original_img,style_img,file_path=file_path):    
    fig=pl.figure(figsize=(6,4))
    ax=fig.add_subplot(121)
    str1='Shape of the original image: %s'
    pl.title(str1%str(original_img.shape),fontsize=int(8))
    ax.imshow(np.squeeze(original_img))
    ax=fig.add_subplot(122)
    str2='Shape of the style image: %s'
    pl.title(str2%str(style_img.shape),fontsize=int(8))
    ax.imshow(np.squeeze(style_img))
    pl.tight_layout(); pl.show()
</script></div><br/>
<div class='linked'><script type='text/x-sage'>
original,style='01_004.png','06_001.png'
for f in [original,style]: get_file(f)
original_img=load_img(original)
style_img=load_img(style)
display_images(original_img,style_img)
</script></div><br/>
     <h2>‚úíÔ∏è &nbsp; Mixed Styles</h2>      
<div class='linked'><script type='text/x-sage'>
hub_model=hub.load(tfhub_path)
stylized_img=hub_model(
    tf.constant(original_img),tf.constant(style_img))[int(0)]
tensor2img(stylized_img)
</script></div><br/>       
  </body>
</html>
```
