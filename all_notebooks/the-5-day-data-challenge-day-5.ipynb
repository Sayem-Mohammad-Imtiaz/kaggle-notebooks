Of the bites, how many were female and how many were male? Let's start by dropping all the NaN values.

```python
#Drop columns..
gender = bites_df.drop(columns = ['vaccination_yrs','vaccination_date','victim_zip','AdvIssuedYNDesc','BreedIDDesc','color','WhereBittenIDDesc','quarantine_date','DispositionIDDesc','head_sent_date','release_date','ResultsIDDesc'])
gender
```

```python
#gender = [gender[i] for i in range(0,len(gender)) if gender[i] != 'UNKNOWN' or gender[i] != 'nan']
gender = gender.dropna()
```

```python
#let's just plot it using seaborn's sns.countplot()
data = gender['GenderIDDesc']
sns.countplot(data)
```

As we can see,  there were about **twice as many male bites reported than female ones.** Males are much more aggressive. 

So. Is this result just a random occurence or is it really always true that males do indeed tend to bite more than females when made to be aggressive? We will perform a chi-square test to solve this case.

```python
gender_ID = gender['GenderIDDesc']
#REMOVE UNKNOWNS:
gender_ID = [gender_ID[i] for i in list(gender_ID.index) if gender_ID[i] != 'UNKNOWN']
```

CREATE A NEW DATAFRAME: 

```python
gender_ID_df = pd.DataFrame(gender_ID, columns = ['Gender ID'])
gender_ID_df.head(10)
```

```python
data = gender_ID_df['Gender ID'].value_counts()
data
```

```python
s.chisquare(data)
```

A large statistic means there is huge difference the effect gender has on the chances that there'll be a bite report. Also the p-value well below the threshold really confirms our hypothesis.
