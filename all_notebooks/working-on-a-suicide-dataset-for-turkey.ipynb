Firstly, we retrieve the data to use it.

```python
data = pd.read_csv("../input/who_suicide_statistics.csv")
```

Then, we put the data in csv to a dataframe. After that, we check it whether it's done.

```python
mydf = pd.DataFrame (data)
print(mydf)
```

Then, we want to see statistical data of the dataset.

```python
mydf.describe()
```

We reach the data of suicides numbers of male which are 15-24 years old in Turkey, then we'll find total number of suicides in the related range.

```python
filt1 = mydf.country == "Turkey"
filt2 = mydf.sex == "male"
filt3 = mydf.age == "15-24 years"
filt4 = mydf.sex == "female"
reqdata = mydf[filt1 & filt2]
mydata = mydf[filt1 & filt2 & filt3]
print(mydata)
summ=0
for each in mydata.suicides_no:
    summ+=each
print("Total suicide numbers of 15-24 years for male in Turkey=", summ)
```

Then, we visualize suicides according to age in Turkey. Group of 35-54 years is attempting to suicide, by statistics.

```python
import matplotlib.pyplot as plt 
countr = mydf [ mydf.country == "Turkey"]
plt.plot(countr.age, countr.suicides_no, color="red", label = "Suicide Numbers", alpha = 1.5)
plt.xlabel("age")
plt.ylabel("suicides_no")
plt.legend()
plt.show()

```

We are looking for suicide numbers of men by ages, in Turkey.  Group of 35-54 years of men is attempting to suicide, by statistics.

```python
plt.plot(reqdata.age, reqdata.suicides_no, color="red", label = "Suicide Numbers", alpha = 1.5)
plt.legend()
plt.xlabel("age")
plt.ylabel("suicides_no")
plt.show()
```

We are looking for suicide numbers of women by ages, in Turkey.  Group of 15-24 years is attempting to suicide, by statistics.

```python
req2 = mydf[filt1 & filt4]
plt.plot(req2.age, req2.suicides_no, color="red", label = "Suicide Numbers", alpha = 1.5)
plt.legend()
plt.xlabel("age")
plt.ylabel("suicides_no")
plt.show()
```

We are evaluating suicides according to sex. Men are attempting to suicide more than womens in Turkey.

```python
req2 = mydf[filt1 & filt4]
plt.scatter(req2.age, req2.suicides_no, color="red", label = "Women Suicide Numbers", alpha = 0.5)
plt.scatter(reqdata.age, reqdata.suicides_no, color="green", label = "Men Suicide Numbers", alpha = 0.2)
plt.legend()
plt.xlabel("sex")
plt.ylabel("suicides_no")
plt.show()
```
