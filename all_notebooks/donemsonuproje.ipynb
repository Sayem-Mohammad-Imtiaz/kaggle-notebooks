aralık belirleme

```python
def SinirlariYazdir(Q1,Q2,Q3,Q4,IQR):
    print("Q1 : ", Q1)
    print("Q3 : ", Q3)
    print("Q2 : ", Q2)
    print("Q4 : ", Q4)
    print("IQR : ", IQR)
    altsinir = Q1 - 1.5 * IQR
    ustsinir = Q3 + 1.5 * IQR
    print("Alt sınır: Q1 - 1.5 * IQR: ", altsinir)
    print("Üst sınır: Q3 + 1.5 * IQR: ", ustsinir)
```

```python
def SinirlariBul(oznitelik):
    outliers= []
    Q1 = df[oznitelik].quantile(0.25)
    Q2 = df[oznitelik].quantile(0.5)
    Q3 = df[oznitelik].quantile(0.75)
    Q4 = df[oznitelik].quantile(1)
    IQR = Q3 - Q1
    print("{0} özniteliği için genel değerler:  ".format(oznitelik))
    SinirlariYazdir(Q1,Q2,Q3,Q4,IQR)
```

### Kategorik degerlerin donusturulmesi

```python
def kategorik_degerleri_duzenle(deger):
    if deger == "Var":
        return 1
    else:
        return 0
```

```python
df["Bluetooth"] = df["Bluetooth"].apply(kategorik_degerleri_duzenle)
df["CiftHat"] = df["CiftHat"].apply(kategorik_degerleri_duzenle)
df["4G"] = df["4G"].apply(kategorik_degerleri_duzenle)
df["3G"] = df["3G"].apply(kategorik_degerleri_duzenle)
df["Dokunmatik"] = df["Dokunmatik"].apply(kategorik_degerleri_duzenle)
df["WiFi"] = df["WiFi"].apply(kategorik_degerleri_duzenle)
```

```python
df.head()
```

```python
def fiyat_degerleri_duzenle(deger):
    if deger == "Çok Ucuz":
        return 0
    elif deger == "Ucuz":
        return 1
    elif deger == "Normal":
        return 2
    else:
        return 3
```

```python
df["FiyatAraligi"] = df["FiyatAraligi"].apply(fiyat_degerleri_duzenle)
```
