## Explorer Breweries in US

### Introduction

This project is intended to understand more about the craft beer business in US based on the craft beer datasets downloaded from Kaggle. Here are the questions I'm interested in answering using statistical analysis and data visualization tools:

1) Which state has the highest number of breweries?
2) Which city has the highest number of breweries? 
3) Which is the most popular beer style in US?
4) What is the average alcohol content? Which beer has the highest alcohol content? 


### Data import and wrangling

```python
import pandas as pd
```

```python
beer_data=pd.read_csv('../input/beers.csv',sep=',')
```

```python
# Replace NaN values with 0 in ibu (bitterness)
beer_data=beer_data.fillna(value=0)
```

```python
brewery_data=pd.read_csv('../input/breweries.csv',sep=',')
```

### Data Exploration 

#### Which states have the highest number of breweries?

```python
states_breweries=brewery_data.groupby(['state']).count()
states_breweries_top=states_breweries.sort(columns='name', axis=0, ascending=False)
states_breweries_top.head()
```

#### Which cities have the highest number of breweries?

```python
cities_breweries=brewery_data.groupby(['city']).count()
cities_breweries_top=cities_breweries.sort(columns='name', axis=0, ascending=False)
cities_breweries_top.head()
```

#### What are the most popular craft beer styles in US?

```python
beer_style=beer_data.groupby(['style']).count()
beer_style_top=beer_style.sort(columns='id', axis=0, ascending=False)
beer_style_top.head()
```

#### What is the average alcohol content in surveryed craft beers? What are the brands with the highest alcohol content?

```python
beer_mean=beer_data['abv'].mean(axis=0)
beer_mean
```

```python
beer_median=beer_data['abv'].median(axis=0)
beer_median
```

```python
beer_abv_top=beer_data.sort(columns='abv', axis=0, ascending=False)
beer_abv_top.head()
```

### Conclusion

-Suprisingly, Colorado has the highest number of breweries among all states followed by California, Michigan, Oregon and Texas.

-Portland has the highest number of breweries among all major US cities followed by Chicago (cheers!), Seattle, Boulder and Austin.

-Apparently IPA is the most popular beer style in US.

-Finally "Quadrupel" has the highest alcohol content, way above the average alcohol content of 5.6%. Make sure you pay extra attention when you drink it :).
