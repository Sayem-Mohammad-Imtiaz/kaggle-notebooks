### It's very famous graph that show relation GDP and Life satifaction.
### I will try this graph by my style visualization !

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv("../input/world-happiness-report-2021/world-happiness-report-2021.csv")
data = data[['Country name','Ladder score','Logged GDP per capita']]
```

### I just make scatterplot that x mean stisfication score and y mean GDP per capita
### But, I think this basic scatter plot's scatter size is so small. So, I will change size and set xlabel, ylabel

```python
plt.figure(figsize=(20,13))
sns.scatterplot(data= data, x='Ladder score',y='Logged GDP per capita')
plt.show()
```

```python
plt.figure(figsize=(20,13))
sns.scatterplot(data= data, x='Ladder score',y='Logged GDP per capita',s=100)
plt.title('Life Satisfication vs GDP per Captia',fontsize=20)
plt.xlabel('Life Satisfication',fontsize=16)
plt.ylabel('GDP per Captia',fontsize=16)
plt.show()
```

### And I want to know countries's name. So I will put text.

```python
plt.figure(figsize=(20,13))
sns.scatterplot(data= data, x='Ladder score',y='Logged GDP per capita',s=100)
plt.title('Life Satisfication vs GDP per Captia',fontsize=20)
plt.xlabel('Life Satisfication',fontsize=16)
plt.ylabel('GDP per Captia',fontsize=16)
for i in range(len(data)):
    plt.text(s =data.loc[i,'Country name'], x=data.loc[i,'Ladder score']+0.01,y=data.loc[i,'Logged GDP per capita']+0.01, fontsize=10)
plt.show()
```

### After put text, It's hard to see because the texts overlap. So I will reduce the scatter's number.

```python
data = data[:50]
plt.figure(figsize=(20,13))
sns.scatterplot(data= data, x='Ladder score',y='Logged GDP per capita',s=100)
plt.title('Life Satisfication vs GDP per Captia',fontsize=20)
plt.xlabel('Life Satisfication',fontsize=16)
plt.ylabel('GDP per Captia',fontsize=16)
for i in range(len(data)):
    plt.text(s =data.loc[i,'Country name'], x=data.loc[i,'Ladder score']+0.01,y=data.loc[i,'Logged GDP per capita']+0.01, fontsize=10)
plt.show()
```

```python
x = np.array(data['Ladder score'])
for i in range(len(x)):
    fit = -0.0747281 *x*x*x + 1.05045076*x*x -3.16337971*x +7.10682006
plt.figure(figsize=(20,13))
sns.scatterplot(data= data, x='Ladder score',y='Logged GDP per capita',s=100)
plt.title('Life Satisfication vs GDP per Captia',fontsize=20)
plt.xlabel('Life Satisfication',fontsize=16)
plt.ylabel('GDP per Captia',fontsize=16)
plt.plot(x,fit,lw=8,alpha=0.3)
for i in range(len(data)):
    plt.text(s =data.loc[i,'Country name'], x=data.loc[i,'Ladder score']+0.01,y=data.loc[i,'Logged GDP per capita']+0.01, fontsize=10)
plt.show()
```

### It's more better than before. But I want to meaningful visualization. So I will plus trend Line

```python
x = np.array(data['Ladder score'])
for i in range(len(x)):
    fit = -0.0747281 *x*x*x + 1.05045076*x*x -3.16337971*x +7.10682006
bound_country = ['Kuwait','Guatemala','Jamaica','Costa Rica','Kosovo','Brazil','El Salvador','Uzbekistan','United Arab Emirates','Singapore','Ireland','Luxembourg']
plt.figure(figsize=(20,13))
sns.scatterplot(data= data, x='Ladder score',y='Logged GDP per capita',s=100)
plt.title('Life Satisfication vs GDP per Captia',fontsize=20)
plt.xlabel('Life Satisfication',fontsize=16)
plt.ylabel('GDP per Captia',fontsize=16)
plt.plot(x,fit,lw=8,alpha=0.3)
for i in range(len(data)):
    if data.loc[i,'Country name'] in bound_country:
        plt.text(s =data.loc[i,'Country name'], x=data.loc[i,'Ladder score']+0.01,y=data.loc[i,'Logged GDP per capita']+0.01, fontsize=10,color='lightgrey')
        plt.scatter(x=data.loc[i,'Ladder score'],y=data.loc[i,'Logged GDP per capita'],color ='lightgrey',s=100)
    else:
        plt.text(s =data.loc[i,'Country name'], x=data.loc[i,'Ladder score']+0.01,y=data.loc[i,'Logged GDP per capita']+0.01, fontsize=10)
plt.show()
```

### I think it tells that Life Satisfication and GDP have relationship. And Change color for ART!

```python
x = np.array(data['Ladder score'])
for i in range(len(x)):
    fit = -0.0747281 *x*x*x + 1.05045076*x*x -3.16337971*x +7.10682006
bound_country = ['Kuwait','Guatemala','Jamaica','Costa Rica','Kosovo','Brazil','El Salvador','Uzbekistan','United Arab Emirates','Singapore','Ireland','Luxembourg']
ax = plt.figure(figsize=(20,13))
ax = plt.axes()
ax.set(facecolor = "whitesmoke")
sns.scatterplot(data= data, x='Ladder score',y='Logged GDP per capita',s=100,color = 'dimgray')
plt.title('Life Satisfication vs GDP per Captia',fontsize=20)
plt.xlabel('Life Satisfication',fontsize=16)
plt.ylabel('GDP per Captia',fontsize=16)
plt.plot(x,fit,lw=8,alpha=0.3)
for i in range(len(data)):
    if data.loc[i,'Country name'] in bound_country:
        plt.text(s =data.loc[i,'Country name'], x=data.loc[i,'Ladder score']+0.01,y=data.loc[i,'Logged GDP per capita']+0.01, fontsize=10,color='gainsboro')
        plt.scatter(x=data.loc[i,'Ladder score'],y=data.loc[i,'Logged GDP per capita'],color ='gainsboro',s=100)
    else:
        plt.text(s =data.loc[i,'Country name'], x=data.loc[i,'Ladder score']+0.01,y=data.loc[i,'Logged GDP per capita']+0.01, fontsize=10,color = 'dimgray')
plt.show()
```
