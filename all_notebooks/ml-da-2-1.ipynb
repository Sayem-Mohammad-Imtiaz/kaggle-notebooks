```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
print('Classes to predict: ', data.target_names)
print('Classes to predict: ', data.feature_names)
X = data.data
y = data.target
print('Number of examples in the data:', X.shape)
print(X[:4])
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
clf = DecisionTreeClassifier(random_state=7)

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=7, train_size=0.8)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import precision_score

print("precision score of classifier : ",precision_score(y_true=y_test, y_pred=y_pred) * 100)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(y_true=y_test, y_pred=y_pred)
print(clf_report)
#we can see from out report, that the recall score for class 1 = 0.96 which is our sensitivity
# and the recall score for class 0 = 0.82 which is our specificity in our case
```
