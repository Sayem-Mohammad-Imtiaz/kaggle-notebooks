```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

df = pd.read_csv('/kaggle/input/weatherww2/Summary of Weather.csv', low_memory=False)

fig = plt.figure()
fig.suptitle('Mean Temperature during WW2', fontsize=14, fontweight='bold')

ax = fig.add_subplot(111)
ax.boxplot(df['MeanTemp'])
#plt.boxplot(df['MeanTemp'])
ax.set_title('Weather data from NOAA')
ax.set_ylabel('Temp °C')
plt.show()
```

```python
fig1 = plt.figure()
fig1.suptitle('Mean Temperature during WW2', fontsize=14, fontweight='bold')
hist = fig1.add_subplot(111)
hist.hist(df['MeanTemp'], orientation='horizontal', color=["purple"])

hist.set_title('Weather data from NOAA')
hist.set_ylabel('Temp °C')
plt.show()
```
