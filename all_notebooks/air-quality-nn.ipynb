1. Data Reduction<br>
The dataset have 15 columns and 9357 rows. I choose 12 columns for data mining. I'm not using date, time, and C6H6(GT) attributes. Whereas C6H6(GT) haven't sensor measure result, since it's not relevant with another attributes that have actual and sensor measures. But, this is just my opinion.<br>
2. Data Cleaning<br>
There are some missing values in this dataset, there are 200 and -200 values in each column, that's missing value. This missing value can make a bad classify result because the value sometimes not relevant with range in some columns. I fill the missing value with mean or median value from each columns. But, first i check if it's a null value.

```python
#reduce columns
df = df.drop('Date', axis=1)
df = df.drop('Time', axis=1)
df = df.drop('C6H6(GT)', axis=1)
```

```python
#check null value
df.isnull().sum()
```

```python
#check minimum values each attribute
df.describe()
```

If you run one by one minimum value will be -200. I do some cross check that every attributes imposibble have minus value except T (Temperature). But, it's impossible too if the Temperature -200, because this case about Very High, High pollution in the air. <br>
But first, i fill this missing value with 0 value, because if i fill with mean, the value was not representative because it's contain with -200 value.

```python
for i in df.ix[df['CO(GT)']<0].index:
    df.loc[i, 'CO(GT)'] = 0
for i in df.ix[df['PT08.S1(CO)']<0].index:
    df.loc[i, 'PT08.S1(CO)'] = 0
for i in df.ix[df['NMHC(GT)']<0].index:
    df.loc[i, 'NMHC(GT)'] = 0
for i in df.ix[df['PT08.S2(NMHC)']<0].index:
    df.loc[i, 'PT08.S2(NMHC)'] = 0
for i in df.ix[df['NOx(GT)']<0].index:
    df.loc[i, 'NOx(GT)'] = 0
for i in df.ix[df['PT08.S3(NOx)']<0].index:
    df.loc[i, 'PT08.S3(NOx)'] = 0
for i in df.ix[df['NO2(GT)']<0].index:
    df.loc[i, 'NO2(GT)'] = 0
for i in df.ix[df['PT08.S4(NO2)']<0].index:
    df.loc[i, 'PT08.S4(NO2)'] = 0
for i in df.ix[df['PT08.S5(O3)']<0].index:
    df.loc[i, 'PT08.S5(O3)'] = 0
for i in df.ix[df['RH']<0].index:
    df.loc[i, 'RH'] = 0
for i in df.ix[df['AH']<0].index:
    df.loc[i, 'AH'] = 0
#this case because it's possible the Temperature have minus degree,
#so i just fill 0 that have -200 value
for i in df.ix[df['T']==-200].index:
    df.loc[i, 'T'] = 0
```

You can see that NMHC_GT median value is 0, that's make posibble if NMHC_GT value contains many 0 values. so that's why i fill the missing value with mean value. But, in some cases like PT08_S4_NO2 and PT08_S5_O3 that have difference quite large between mean and median, possibility because both of this column have a lot of pretty high value. So, i choose to fill both of this attributes with median value.

```python
for i in df.ix[df['CO(GT)']==0].index:
    df.loc[i, 'CO(GT)'] = df['CO(GT)'].mean()
for i in df.ix[df['PT08.S1(CO)']==0].index:
    df.loc[i, 'PT08.S1(CO)'] = df['PT08.S1(CO)'].mean()
for i in df.ix[df['NMHC(GT)']==0].index:
    df.loc[i, 'NMHC(GT)'] = df['NMHC(GT)'].mean()
for i in df.ix[df['PT08.S2(NMHC)']==0].index:
    df.loc[i, 'PT08.S2(NMHC)'] = df['PT08.S2(NMHC)'].mean()
for i in df.ix[df['NOx(GT)']==0].index:
    df.loc[i, 'NOx(GT)'] = df['NOx(GT)'].median()
for i in df.ix[df['PT08.S3(NOx)']==0].index:
    df.loc[i, 'PT08.S3(NOx)'] = df['PT08.S3(NOx)'].mean()
for i in df.ix[df['NO2(GT)']==0].index:
    df.loc[i, 'NO2(GT)'] = df['NO2(GT)'].mean()
for i in df.ix[df['PT08.S4(NO2)']==0].index:
    df.loc[i, 'PT08.S4(NO2)'] = df['PT08.S4(NO2)'].median()
for i in df.ix[df['PT08.S5(O3)']==0].index:
    df.loc[i, 'PT08.S5(O3)'] = df['PT08.S5(O3)'].median()
for i in df.ix[df['T']==0].index:
    df.loc[i, 'T'] = df['T'].mean()
for i in df.ix[df['RH']==0].index:
    df.loc[i, 'RH'] = df['RH'].mean()
for i in df.ix[df['AH']==0].index:
    df.loc[i, 'AH'] = df['AH'].mean()
```

This dataset dont have class. So i use k-means clustering to fill class because this clustering algorithm very simple and have pretty good accuracy. I choose 2 clusters because in data description describe that the sensor was located on the field in a significantly polluted area. So cluster 0 represent very highly polluted, and cluster 1 represent highly polluted.

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=2, random_state=1)
new = df._get_numeric_data().dropna(axis=1)
km.fit(new)
predict=km.predict(new)
```

```python
#fill dataframe with cluster result
df['Class'] = pd.Series(predict, index=df.index)
```

```python
df.head()
```

```python
X = df.drop('Class', axis=1)
y = df.loc[:,'Class'].values
```

```python
from sklearn.cross_validation import train_test_split
#split data with 70%, 30%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,  random_state =0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
#use nn for classify
from sklearn.linear_model import Perceptron
ppn =  Perceptron(n_iter=100, eta0=0.1, random_state=0)
ppn.fit(X_train_std, y_train)
```

```python
y_pred = ppn.predict(X_test_std)
print('Misclassified sample : %d' % (y_test != y_pred).sum())

from sklearn.metrics import accuracy_score
print('accuracy : %.2f' % accuracy_score(y_test, y_pred))
```

It is very high accuracy, maybe because this dataset dont have actual class. I get advice that maybe this is called overfitting.
