```python
import pandas as pd
df1 = pd.read_csv("../input/fromkry2/task1_products.csv")
df2 = pd.read_csv("../input/fromkry/task1_currencies.csv")
display(df1.info())
display(df2.info())
```

```python
df1.head()
```

```python
df1.columns = ["price"]
```

```python
df1 = df1.reset_index()
```

```python
df1["price"] = df1["index"].apply(lambda x: x.split(";")[-1])
                               
df1.price = df1.price.astype("float")
```

```python
for col in df2.currency.unique():
    df1[col] = df2[df2.currency==col].currency_rate.values[0]
```

```python
df1
```

```python
for col in df2.currency.unique():
    df1[col] *= df1.price
```
