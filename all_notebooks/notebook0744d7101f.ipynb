```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


rd = pd.read_csv('/kaggle/input/league-of-legends-diamond-ranked-games-10-min/high_diamond_ranked_10min.csv')

rd.head()
```

```python
blue_statistic = rd.loc[:,'blueWardsPlaced': 'blueGoldPerMin']
red_statistic = rd.loc[:, 'redWardsPlaced':]
label = rd['blueWins']

differentiate = blue_statistic.subtract(red_statistic.values, axis = 0)

datasets = differentiate.join(label)
```

```python
datasets = datasets.sample(frac=1, random_state=42).reset_index(drop=True)

label = datasets['blueWins']
del datasets['blueWins']
datasets.head()
```

```python
train_set, test_set, val_set = np.split(datasets,[int(0.6 * len(datasets)), int(0.8 * len(datasets))])
train_label, test_label, val_label = np.split(label,[int(0.6 * len(label)), int(0.8 * len(label))])
```

```python
train_set = train_set.to_numpy()
test_set = test_set.to_numpy()
```

```python
x_train = train_set
x_test = test_set

y_train = train_label
y_test = test_label
```

```python
from sklearn.ensemble import RandomForestClassifier
# n_estimators = number of decision trees
rf = RandomForestClassifier(n_estimators=30, max_depth=9)
rf.fit(x_train, y_train)
print("score on test: " + str(rf.score(x_test, y_test)))
print("score on train: "+ str(rf.score(x_train, y_train)))
```
