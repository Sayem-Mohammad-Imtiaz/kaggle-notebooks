```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
```

```python
data_csv = pd.read_csv("../input/ligue-1-saison-20202021/data_2020_2021.csv")
#data_csv
```

Data Processing

```python
data_processed = data_csv.loc[data_csv['home_goal']!="-"]
#data_processed
```

Calculate Total Goals per Game

```python
data = data_processed.assign(total_goal = pd.to_numeric(data_processed["home_goal"]) + pd.to_numeric(data_processed["away_goal"]))
data
```

Data Analysis

```python
total_goals = data["total_goal"]
print("Nombre de buts moyens par match dans la saison")
total_goals.mean()
```

```python
home_goals = data[["home", "total_goal"]].groupby(by="home").sum()
#home_goals
#home_goals.sort_values(by=["total_goal"],ascending = False)
```

```python
away_goals = data[["away", "total_goal"]].groupby(by="away").sum()
#away_goals
#away_goals.sort_values(by=["total_goal"],ascending = False)
```

```python
total_goals = home_goals+away_goals
home_count = 2*data[["home", "total_goal"]].groupby(by="home").count()
total_goals = total_goals/home_count
print("Nombre de buts par matchs par equipe")
total_goals.sort_values(by="total_goal",ascending=False) 
```

Estimations and Probabilities

```python
data_sup15 = data.loc[data['total_goal']>1.5]
print("% de match avec + de 1.5 buts dans la saison :", 100*len(data_sup15)/len(data))
```

```python
home_goals = data_sup15[["home", "home_goal"] ].groupby(by="home").count()
home_games = data[["home", "home_goal"] ].groupby(by="home").count()
away_goals = data_sup15[["away", "away_goal"] ].groupby(by="away").count()
away_games = data[["away", "away_goal"] ].groupby(by="away").count()

average_goal_15 = 100*home_goals/home_games
average_goal_15 = average_goal_15.assign(away = 100*away_goals/away_games)
print("probabilite d'avoir plus de 1.5+ buts dans le match en fonction de l'equipe")
average_goal_15
```

Dans 90% des cas, lorsque des equipes comme Marseille, Monaco ou Brest jouent a domicile les matchs finissent avec au moins deux buts. A l'inverse des equipes comme Metz ou Rennes creent des matchs fermes lorsqu'ils se deplacent a l'exterieur (68% des matchs ont au moins deux buts).

```python
print("Moyenne des probabilites des matchs a + de 1.5 buts")
average_goal_15["home_goal"].mean()
```

En dessous de 78.1 on peut considerer que c'est une equipe qui propose des matchs plutot fermes. Au dessus, l'equipe proposera plutot des matchs ouverts.
