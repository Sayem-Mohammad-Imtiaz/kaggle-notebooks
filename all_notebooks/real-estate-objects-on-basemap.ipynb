Hi, this is visulization of Russia Real Estate Objects 2018-2021 for better understanding.

## 1. Loading Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

## 2. Loading Data

```python
df = pd.read_csv("/kaggle/input/russia-real-estate-20182021/all.csv")
df.sample(10)
```

## Plot on Basemap

```python
# read in data to use for plotted points
lat = df['geo_lat'].values
lon = df['geo_lon'].values

# Prepare basemap

plt.figure(figsize=(20,15))

m = Basemap(projection="mill"
            ,llcrnrlat=lat.min()-5
            ,urcrnrlat=lat.max()+5
            ,llcrnrlon=lon.min()-5
            ,urcrnrlon=lon.max()+5
            ,resolution='c'
            ,epsg=4269)
m.drawlsmask(land_color='coral',ocean_color='aqua',lakes=True)
m.drawcoastlines(linewidth=0.5)
m.drawcountries(linewidth=1.5) 


parallels = np.arange(0.,81,10.)
meridians = np.arange(0.,360.,10.)

# Pickup coordinates
m.plot(lon, lat, 'ro', markersize=2 ,alpha=.05)

plt.title("Russia Real Estate Objects 2018-2021")
plt.show()
```

We can find out most of Russia Real Estate Objects located on west, and looks like 4 groups : 

**1.  Moscow - Tyumen**

**2.  Near by Black Sea**

**3.  St.Petersburg**

**4.  Omak - Krasnoyarsk**
