```python
import requests

URL = 'https://www.medicineindia.org/hospitals-in-india'
page = requests.get(URL)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
table = soup.find(id='main')
```

```python
# Finding container class with id
print(table.prettify())
```

```python
# finding individual classes with classsname
hospitals = table.find_all('tr', class_='hospital-row')

#print few
for i in hospitals[:5]:
  print(i)
```

```python
print(hospitals[0].prettify())
```

```python
import pandas as pd
```

```python
l1=[]
l2=[]
l3=[]
for i in hospitals:
  l = [j.text for j in i]
  l1.append(l[0])
  l2.append(l[1])
  l3.append(l[2])
```

```python
df = pd.DataFrame({'Hospital':l1, 'State':l2, 'City':l3})
```

# **Landing Page scraping done**
Now we need to dive into each hospital page and gather some more info

```python
#Getting a list of all urls with previously created object

urls = []
for i in hospitals:
  # print(i.prettify())
  urls.append("http://" + i.find_all('a')[0]['href'][2:])
```

```python
pincode = []
local_address = []
for url in urls:
  page = requests.get(url)
  soup = BeautifulSoup(page.content, 'html.parser')
  details = soup.find(id='main')
  # print(details.prettify())
  pin = (details.find_all('dd', itemprop='postalCode'))[0]
  address = (details.find_all('dd', itemprop='streetAddress'))[0]
  pincode.append(pin.text)
  local_address.append(address.text)
```

```python
len(local_address)
```

```python
df['LocalAddress'] = pd.Series(local_address)
df['Pincode'] = pd.Series(pincode)
df.head()
```

```python
df.to_csv('./HospitalsInIndia.csv')
```
