# Import Dataset
Temperatures dataset imported the to the workspace.

```python
df = pd.read_csv('../input/temperature-change/Environment_Temperature_change_E_All_Data_NOFLAG.csv', encoding='latin-1')
df.head(5)
```

# Dataset Editing
Dataset headers have been renamed. Inappropriate data cleared in the months column.

```python
df=df.drop(columns=['Area Code','Months Code','Element Code','Unit'])
df=df.loc[df.Months.isin(['January', 'February', 'March', 'April', 'May', 'June', 'July','August', 'September', 'October', 'November', 'December'])]
df
```

```python
df.isna().sum()
```

# Turkey's Data Selected
Only in Turkey's data was selected data sets.

```python
Tr_df=df[df["Area"]=="Turkey"]
Tr_df
```

```python
Tr_df=Tr_df.melt(id_vars=['Area','Months','Element'],var_name='Year', value_name='Temperature')
Tr_df['Year']=Tr_df['Year'].str[1:].astype('str')
Tr_df
```

```python
Tr_sd=Tr_df[Tr_df["Element"]=="Standard Deviation"]
Tr_sd.describe()
```

```python
Tr_tc=Tr_df[Tr_df["Element"]=="Temperature change"]
Tr_tc.describe()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(Tr_tc.Temperature.loc[Tr_tc.Element=='Temperature change'],kde=True,stat='density')
plt.axvline(x=0.0, color='b', linestyle='-')
plt.xlabel('temperature change')
plt.title('temp change in Turkey')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(Tr_tc['Year'].loc[Tr_tc.Element=='Temperature change'],Tr_tc['Temperature'].loc[Tr_tc.Element=='Temperature change'])
plt.plot(Tr_tc.loc[Tr_tc.Element=='Temperature change'].groupby(['Year']).mean(),'r',label='Average')
plt.axhline(y=0.0, color='b', linestyle='-')
plt.xlabel('Year')
plt.xticks(np.linspace(0,58,20),rotation=45)
plt.ylabel('Temperature change')
plt.legend()
plt.title('temp change in Turkey')
plt.show()
```
