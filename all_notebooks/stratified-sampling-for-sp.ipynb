```python
obj_df = df.select_dtypes(include=['object']).copy()
obj_df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
obj_df["gender_code"] = ord_enc.fit_transform(obj_df[["gender"]])
#obj_df[["gender", "gender_code"]].head(11)

ord_enc = OrdinalEncoder()
obj_df["ever_married_code"] = ord_enc.fit_transform(obj_df[["ever_married"]])
#obj_df[["ever_married", "ever_married_code"]].head(11)

ord_enc = OrdinalEncoder()
obj_df["work_type_code"] = ord_enc.fit_transform(obj_df[["work_type"]])

ord_enc = OrdinalEncoder()
obj_df["Residence_type_code"] = ord_enc.fit_transform(obj_df[["Residence_type"]])

ord_enc = OrdinalEncoder()
obj_df["smoking_status_code"] = ord_enc.fit_transform(obj_df[["smoking_status"]])
```

```python
obj_df.iloc[:,5:11]
```

```python
df=df.drop("gender" ,axis=1)
df=df.drop("ever_married" ,axis=1)
df=df.drop("work_type" ,axis=1)
df=df.drop("Residence_type" ,axis=1)
df=df.drop("smoking_status" ,axis=1)
```

```python
df["gender"] = obj_df.iloc[:,5:6]
df["ever_married"] = obj_df.iloc[:,6:7]
df["work_type"] = obj_df.iloc[:,7:8]
df["Residence_type"] = obj_df.iloc[:,8:9]
df["smoking_status"] = obj_df.iloc[:,9:10]
df =df.dropna()
```

```python
df.iloc[:,1:13].corr()
```

```python
import matplotlib.pyplot as plt


labels =df['stroke'].value_counts(sort = True).index
sizes = df['stroke'].value_counts(sort = True)

colors = ["lightblue","red"]
explode = (0.05,0) 
 
plt.figure(figsize=(7,7))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90,)

plt.title('Customer Churn Breakdown')
plt.show()
```

```python
X1 = df.iloc[:,1:6] # degiskenler
X2 = df.iloc[:,7:13] # degiskenler
```

```python

X= pd.concat([X1, X2] , axis=1)
X.head()
```

```python
X.info()
```

```python
Y = df.iloc[:,6] # Stroke
Y.head(5)
```

```python
Y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y,
                                                    stratify=Y, 
                                                    test_size=0.2 ,random_state=42)
```

```python
plt.bar(y_train ,y_train.index)
```

```python
plt.bar(y_test ,y_test.index)
```

```python
count = sum(y_test)
count
```

```python
forest = RandomForestClassifier(n_estimators = 200)
#== 
forest.fit(X_train, y_train)
#== 
#Score 
#== 
forest_score = forest.score(X_train, y_train)
forest_test = forest.score(X_test, y_test)
#== 
#testing model 
#== 
y_pred = forest.predict(X_test)
#== 
#evaluation
#== 
cm = confusion_matrix(y_test,y_pred)
print('Training Score',forest_score)
print('Testing Score \n',forest_test)
print(cm)
```

```python
plt.figure(figsize=(14,5))
cm = confusion_matrix(y_test,y_pred)
conf_matrix = pd.DataFrame(data=cm,columns=['Predicted:0','Predicted:1'],index=['Actual:0','Actual:1'])
sns.heatmap(conf_matrix, annot=True,fmt='d',cmap="Reds");
```
