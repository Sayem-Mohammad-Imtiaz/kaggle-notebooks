```python
import pandas as pd
df=pd.read_csv("/kaggle/input/suicide-rates-overview-1985-to-2016/master.csv")
df
```

```python
#importing all the libraries
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
#Rename for ease of use
df.rename(columns={"suicides/100k pop":"suicides_pop","HDI for year":"HDI_for_year",
                  " gdp_for_year ($) ":"gdp_for_year"," gdp_per_capita ($) ":"gdp_per_capita",
                    "gdp_per_capita ($)":"gdp_per_capita"}, inplace=True)
print(df.columns)
```

```python
df["gdp_for_year"] = df["gdp_for_year"].str.replace(",","").astype(np.int64)
df["age"] = df["age"].str.replace("5-14 years","05-14 years")
```

**Line Graph showing the number of suicde rates per year**

```python
df_all=df
sns.lineplot(df_all.year, df.suicides_no, ci = None)
plt.legend(["people"])
plt.show()
```

**Comparing the Numbers Gender wise**

```python
df_men = df[df.sex == "male"]
df_women = df[df.sex == "female"]
sns.lineplot(df_men.year, df.suicides_no, ci = None)
sns.lineplot(df_women.year, df.suicides_no, ci = None)
plt.legend(["male", 'female'])
plt.show()
```

```python
#Pie graph for gender comparison of suicide
t1 = pd.pivot_table(df, values='suicides_no', index=['country'], columns=['sex'], aggfunc=np.sum)
plt.figure(figsize=(8, 7))
splot = t1[['male','female']].sum().plot.pie(autopct='%1.1f%%')
plt.ylabel("Sex")
```

**Checking the suicide rate categorical wise or age wise**

```python
#age wise comparison
df_age = df.groupby(["year","age"])["suicides_no", "population"].sum()
df_reset = df_age.copy().reset_index()
plt.figure(figsize=(9,6))
sns.lineplot("year", df_reset.suicides_no*100/df_reset.population, hue = "age",
             data = df_reset, linewidth = 2.5, style = "age")
plt.xticks(rotation = 90)
plt.show()
```

**We can clearly see from the above data visualization and analysis that the rate of suicide or the number of suicide every year has been generally decreasing but in the recent years the rate has started to increase again. Also we can see that the rate of suicide in male category is more than that of the female category.**
