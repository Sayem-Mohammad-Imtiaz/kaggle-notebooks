```python
#IMPORT DATA ANALYSIS , DATA VISUALIZATION AND GEOGRAPHIC LIBRARIES
import pandas as pd #data analysis
import matplotlib.pyplot as plt #data visualization
import seaborn as sns
import numpy as np #algebra

import plotly.plotly as py #used to plot data on a map
import plotly.graph_objs as go 
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot

d2015=pd.read_csv("../input/world-happiness/2015.csv") #uploading year 2015 file

```

THE GRAPH SHOWS THE HAPPINESS RATE IN DIFFERENT COUNTRIES OF WORLD.
THE COUNTRIES WITH DARKER SHADES OF RED HAVE HIGHER HAPPINESS RATE AS COMPARE TO THE COUNTRIES WITH LIGHTER SHADE OF RED.
THINGS TO NOTE : WEST COUNTRY POSSES DARK SHADES OF RED (MORE HAPPY)
WHERE AS ASIAN AND AFRICAN HAVE LIGHTER SHADES (LESS HAPPY)
 **LETS CHECK IS OUT WHY ??????? **

```python
#HERE I AM CREATING A WORLD MAP WHICH TELLS THE AMOUNT OF HAPPINESS IN DIFFERENT COUNTRIES
#THE DARKER THE SHADES ,THE MORE IS THE HAPPINESS
init_notebook_mode(connected=True) 
data = dict(
        type = 'choropleth',
        locations = d2015['Country'],
        locationmode='country names',   
        z = d2015['Happiness Score'],
        text = d2015['Region'],
        colorbar = {'title' : 'Happiness Score'},
      ) 

layout = dict(
    title = 'Happiness Score',
    geo = dict(
        showframe = False,
        projection = {'type':'Mercator'}
    )
)
choromap1 = go.Figure(data = [data],layout = layout)
iplot(choromap1)
```

after seeing the world map we can consider that countries in **Africa** and **Asian** countries posses a low happiness rate.

```python
#this is dataset on kaggle with name "idependence days" it contains different nation with their year of independence 
independence=pd.read_csv("../input/independence-days/independence-days.csv")
```

```python
independence.head(5)
```

NOW THIS GRAPH SHOWS WHICH COUNTRY HAD GOT INDEPENDENCE IN WHICH YEAR ........ THE COUNTRIES WHICH HAD GOT INDEPENDENCE LATELY ARE SHOWN IN DARKER SHADES OF RED (**ASIAN**,**AFRICAN**) ........ THE COUNTRIES WHICH HAD GOT INDEPENDENCE AT EARLY STAGE ARE SHOWN WHICH LIGHTER SHADE OF RED (**WESTERN COUNTRIES**)

```python
#creating a map for independence years of each country
init_notebook_mode(connected=True) 
data = dict(
        type = 'choropleth',
        locations = independence['Country'],
        locationmode='country names',   
        z = independence['Year celebrated'],
        text = independence['Country'],
        colorbar = {'title' : 'Year of Independence'}
      ) 

layout = dict(
    title = 'Independence Year',
    geo = dict(
        showframe = False,
        projection = {'type':'Mercator'}
    )
)
choromap = go.Figure(data = [data],layout = layout)

iplot(choromap)
```

**COMPARING BOTH MAPS SIDE BY SIDE**

```python
#now let us understand whats going there , till now we have created two map the first world map shows the happiness score and the second world
#map show the year of independence
#the country which got independence earlier are more happier  then the countries whichh got  independence much lately.
iplot(choromap)

iplot(choromap1)
```
