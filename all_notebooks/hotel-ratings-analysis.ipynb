Hi! I will be exploring a dataset dedicated to hotel reviews and their ratings. The ratings can depend on numerous factors, but I specifically wanted to observe correlations between elements that can increase hotel ratings. Let's take a look at the data! :)

```python
#Load data
reviews = pd.read_csv('../input/hotel-reviews-dataset-enriched/hotel_reviews_enriched.csv', usecols=[0,8,9,10,17,36, 37])
#reviews = pd.read_csv('../input/hotel-reviews-dataset-enriched/hotel_reviews_enriched.csv')
reviews.info()
```

```python
reviews.isnull().sum()
```

The dataset was enriched with nearby buisnesses within a 100m, 1km, and 5km radius. This information was joined from a Yelp dataset. I would like to explore if a higher number of close-by buisnesses is assoicated with a higher review score.

Due to the size of the dataset and the fact that we will be using simple linear regression, we will remove the null valued entries

```python
reviews.dropna()
```

A correlation heat map would be an interesting way view correlation between the dependent Y (Reviewer_Score) and the X

```python
f, ax = plt.subplots(figsize=(10, 8))
corr = reviews.corr()
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(240,10,as_cmap=True),
            square=True, ax=ax)
```

```python
f= plt.figure(figsize=(12,5))

ax=f.add_subplot(221)
sns.lineplot(x=reviews["Businesses_100m"], y=reviews["Reviewer_Score"], ax=ax)
ax.set_title('>100m buisnesses on reviewer score')

ax=f.add_subplot(222)
sns.lineplot(x=reviews["Businesses_1km"], y=reviews["Reviewer_Score"], ax=ax)
ax.set_title('>1km buisnesses on reviewer score')

ax=f.add_subplot(223)
sns.lineplot(x=reviews["Businesses_5km"], y=reviews["Reviewer_Score"], ax=ax)
ax.set_title('>5km buisnesses on reviewer score')
```
