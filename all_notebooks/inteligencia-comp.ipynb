```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

df = pd.read_csv("/kaggle/input/brasileiro/brasileiro.csv")
dataset = df.values

for ds in dataset:
    ds[4] = float(str(ds[4]).split("%")[0])/100

inputs = dataset[:,1:9]
output = dataset[:,9]

minMaxScaler = preprocessing.MinMaxScaler()
inputsFormatados = minMaxScaler.fit_transform(inputs)

inputsParaTreinamento, inputsParaTest, outputsParaTreinamento, outputsParaTest = train_test_split(inputsFormatados, output, test_size=0.05)

model = keras.Sequential([
    layers.Dense(32, activation='relu', input_shape=(8,)),
    layers.Dense(32, activation='relu'),
    layers.Dense(1, activation='sigmoid'),
])
model.compile(optimizer='sgd',
              loss='binary_crossentropy',
              metrics=['accuracy'])

inputsParaTreinamento=np.asarray(inputsParaTreinamento).astype(np.int)
outputsParaTreinamento=np.asarray(outputsParaTreinamento).astype(np.int)

hist = model.fit(inputsParaTreinamento, outputsParaTreinamento,
          batch_size=32, epochs=100)


inputsParaTest=np.asarray(inputsParaTest).astype(np.int)
outputsParaTest=np.asarray(outputsParaTest).astype(np.int)
model.evaluate(inputsParaTest, outputsParaTest)

plt.plot(hist.history['loss'])
```
