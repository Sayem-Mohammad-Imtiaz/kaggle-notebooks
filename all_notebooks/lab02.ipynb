```python
def cov(x,y):
    x_m=np.mean(x)
    y_m=np.mean(y)
    s=0
    for i in range(len(x)):
        s+=(x[i]-x_m)*(y[i]-y_m)
    return s/len(x)
        
def cov_mat(textfile_location):
    text_file=open(textfile_location,'r')
    data=list()
    for line in text_file:
        for val in line.split(','):
            data.append(list())
        break
        
    for line in text_file:
        line_data=line.replace('\n', '').split(',')
        for i in range(len(line_data)):
            data[i].append(float(line_data[i]))
            
    n=len(data)     
    cmat=np.zeros((n,n))
    for i in range(n):
        for j in range(i):
            cmat[i][j]=cov(data[i],data[j])
            cmat[j][i]=cmat[i][j]
        cmat[i][i]=cov(data[i],data[i])
    #print(cmat)
    
    
cov_mat('../input/nagraj/example_data.TXT')
```
