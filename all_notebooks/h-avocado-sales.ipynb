```python
df = pd.read_csv("/kaggle/input/avocado-prices/avocado.csv")
df.drop(["Unnamed: 0", "year"], inplace = True, axis =1)
df["Date"] = pd.to_datetime(df["Date"])
df.head()
```

```python
df.info()
```

```python
sns.set(font_scale=1.5)
plt.figure(figsize=(10,10))
sns.scatterplot( "Total Volume", "AveragePrice", data= df[~df["region"].isin( ["TotalUS"])] , hue = "type")
```

```python
df_organic = df[df["type"]== "organic"]
df_organic
```

```python
# plt.figure(figsize=(20,20))
sns.countplot("type", data= df)
# plt.xticks(rotation= 90)
```

```python
df_region = df.groupby("region", as_index=False).mean().sort_values(by= "AveragePrice", ascending = False).reset_index()
df_region.drop("index", axis = 1, inplace = True)
df_region.head()
```

```python
df_type_grouped = df.groupby("type", as_index=False).mean().sort_values(by= "AveragePrice", ascending=False).reset_index()
df_type_grouped
```

```python
print("there are %2d types of avocado, avreage price are %5.2f and %5.2f" \
      %(len(df_type_grouped), df_type_grouped.loc[0, "AveragePrice"], df_type_grouped.loc[1, "AveragePrice"]))
```
