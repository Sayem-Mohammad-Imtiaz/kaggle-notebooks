**4.Analyze, identify patterns, and explore the data.**

 ทำการสำรวจข้อมูล และวิเคราะห์ข้อมูล เปรียบเทียบว่าควรนำข้อมูลอะไรไปใช้ในการทำงาน เพื่อใช้ในการคำนวณ ซึ่งสิ่งที่จำเป็นต้องใช้มีดังนี้ คือ
        4.1 rating
        4.2 vote_count
        4.3 popularity
        4.4 Weighted_rating**

```python
#เลือกข้อมูลที่จำเป็นต้องใช้
movies_df = movies_df[[ 'id','original_title','rating','vote_count','popularity']]
movies_df.head()
```

```python
#เริ่มทำการกำหนดตัวแปลเพื่อใช้ในการคำนวณหาคะแนน 
R = movies_df['rating']
v = movies_df['vote_count']
m = movies_df['vote_count'].quantile(0.7) #กำหนด 30% ของหนังที่ได้รับความนิยม
C = movies_df['rating'].mean()
```

```python
#เราจะได้สูตรในการคำนวณ weighted_rating คือ ((R*v)+(C*m)) / (v+m)
movies_df['weighted_rating'] = ((R*v)+(C*m)) / (v+m)
```

**6.Visualize, report, and present the problem solving steps and final solution.**

 จากการวิเคราะห์ข้อมูล และกรองข้อมูลเรียบร้อยแล้ว เรามาทำการคำนวณหา  Top 20 Movie ที่เราได้จัดอันดับโดยใช้ข้อมูลของ Wighted rating  และ Popularity  ซึ่งข้อมูลส่วนนี้สามารถนำไปใช้ในการตัดสิน ของผู้สร้างภาพยนตร์ หรือ ผู้กำกับภาพยนตร์ว่าจะสร้างหนังอะไร แนวไหน เพื่อที่จะให้ตอบโจทย์ความต้องการของผู้คนที่ชอบดูภาพยนตร์ อีกทั้งสามารถสร้างรายได้มหาศาล ถ้าเรารู้ว่าผู้คนชอบดูหนังแนวไหนโดยข้อมูลจะมีกราฟบอกอย่างชัดเจน 

```python
#จัดการเรื่องทำข้อมูลและ import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler

scalar = MinMaxScaler()
scaled_scores = pd.DataFrame(scalar.fit_transform(movies_df[['popularity','weighted_rating']]), 
                             columns=['popularity_scaled','weighted_rating_scaled'])
movies_scaled = movies_df.copy()
movies_scaled[['popularity_scaled','weighted_rating_scaled']] = scaled_scores
movies_scaled
```

```python

movies_scaled['score'] = (0.5 * movies_scaled['weighted_rating_scaled']) + \
                            (0.5 * movies_scaled['popularity_scaled'])
```

```python
#ทำการจัดเรียงข้อมูลโดยการใช้ข้อมูลของ hybrid_score
hybrid_score=movies_scaled.sort_values('score',ascending=False)
#เป็นการกำหนด ขนาดของกราฟ
plt.figure(figsize=(12,6))
#เป็นการกำหนดตำแหน่งของกราฟให้ title อยู่แกน x แล้วให้ rating อยู่แกน y โดยใช้ข้อมูลของ hybrid_score
axis1=sns.barplot(x=hybrid_score['score'].head(20), y=hybrid_score['original_title'].head(20), data=hybrid_score);
#เป็นการกำหนด ชื่อของกราฟ
plt.title('Best  Hybrid score', weight='bold')
#เป็นการกำหนดการใช้ชื่อตัวแปลของแกน x
plt.xlabel('Hybrid Score', weight='bold')
#เป็นการกำหนดการใช้ชื่อตัวแปลของแกน y
plt.ylabel('Movie Title', weight='bold');
```
