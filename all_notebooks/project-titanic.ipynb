```python
y_train=np.array([x_train[x_train.columns[0]]])

y_train=y_train.T
x_train.drop(x_train.columns[0],inplace=True, axis=1)

y_test=np.array([x_test[x_test.columns[0]]])
x_test.drop(x_test.columns[0],inplace=True, axis=1)
y_test=y_test.T
```

```python
x_train=np.array(x_train)
x_test=np.array(x_test)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
print(x_train[:,:])
```

```python

m=len(x_train)
n=len(x_train[0])
print(m,n)
```

```python
def sig(x):
    return 1/(1+np.exp(-x))
c=0.0000000000000000000000000001
th=np.array([[0.00]*n])

th=th.T
th0=0
alpha=0.00001
j=[]
model=np.dot(x_train,th)+th0
h=sig(model)
for i in range(0,len(h),1):
    if(h[i]>=0.5):
        h[i]=1
    else:
        h[i]=0
temp=0
for i in range(0,m,1):
    temp+=(y_train[i]*np.log(h[i]+c)+(1-y_train[i])*np.log(1-h[i]+c))
temp=-temp/m
print("j is : ",temp)
j.append(temp)
print(j)
```

```python
#gradient descent
for it in range(1,200,1):
    model=np.dot(x_train,th)+th0
    h=sig(model)
    for i in range(0,len(h),1):
        if(h[i]>=0.5):
            h[i]=1
        else:
            h[i]=0
    dth=(1/m)*(np.dot(x_train.T,(h-y_train)))
    dth0=(1/m)*(np.sum(h-y_train))
    th-=(alpha*dth)
    th0=th0-alpha*dth0
    
    model=np.dot(x_train,th)+th0
    h=sig(model)
    for i in range(0,len(h),1):
        if(h[i]>=0.5):
            h[i]=1
        else:
            h[i]=0
    temp=0
    for i in range(0,m,1):
        temp+=(y_train[i]*np.log(h[i]+c)+(1-y_train[i])*np.log(1-h[i]+c))
    temp=-temp/m
    j.append(temp)
  
    
print(th0)
print(th)
plt.figure(0)
plt.xlabel('iterations')
plt.ylabel('J')
plt.plot(j)


#prediction
model=np.dot(x_test,th)+th0
h=sig(model)
for i in range(0,len(h),1):
    if(h[i]>=0.5):
        h[i]=1
    else:
        h[i]=0
plt.figure(1)
plt.plot(y_test,'*')
plt.plot(h,'*')
```

```python
print("accuracy score of model from scratch is : ",accuracy_score(h,y_test)*100)
model2=LogisticRegression()
model2.fit(x_train,y_train)
y_pred=model2.predict(x_test)
print("accuracy score of model with libraries is : ",accuracy_score(y_pred,y_test)*100)
for i in range(0,len(y_test),1):
    print(y_test[i],y_pred[i])
```
