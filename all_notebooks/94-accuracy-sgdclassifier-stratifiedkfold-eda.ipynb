```python
df = df.drop('id',axis=1)
```

```python
df.head()
```

```python
df = pd.get_dummies(columns=['gender','ever_married','work_type','Residence_type','smoking_status'],data=df,drop_first=True)
```

```python
df.head()
```

```python
df.columns
```

```python
X = df[['age', 'hypertension', 'heart_disease', 'avg_glucose_level', 'bmi',
        'gender_Male', 'gender_Other', 'ever_married_Yes',
       'work_type_Never_worked', 'work_type_Private',
       'work_type_Self-employed', 'work_type_children', 'Residence_type_Urban',
       'smoking_status_formerly smoked', 'smoking_status_never smoked',
       'smoking_status_smokes']]
y = df.stroke
```

```python
from sklearn.linear_model import SGDClassifier
sgd_cal = SGDClassifier()
```

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(sgd_cal,X,y,cv=10)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score
```

```python
score.mean()
```

```python
X.shape,y.shape
```

```python
df.stroke.value_counts()
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
accuracy = []
skfl = StratifiedKFold(n_splits=10,random_state=None)
skfl.get_n_splits(X,y)
for train_index,test_index in skfl.split(X,y):
    print('train:',train_index,'validation:',test_index)
    X1_train , X1_test = X.iloc[train_index] , X.iloc[test_index]
    y1_train , y1_test = y.iloc[train_index] , y.iloc[test_index]
    
    sgd_cal.fit(X1_train,y1_train)
    prediction = sgd_cal.predict(X1_test)
    score = accuracy_score(prediction,y1_test)
    accuracy.append(score)
    
print(accuracy)
```

```python
np.array(accuracy).mean()

```
