```python
#Deleting unncecessary Columns
zomato=zomato_real.drop(['Web_Link', 'Phone_number'], axis=1) #Dropping the column "Web_Link', 'Phone_number" and saving the new dataset as "zomato"
```

```python
#Remove the NaN values from the dataset
zomato.isnull().sum()
zomato.dropna(how='any', inplace=True)
zomato.info() #.info() function is used to get a concise summary of the dataframe
```

```python
#Reading the Column Names
zomato.columns
```

```python
#Changing Column Names
zomato = zomato.rename(columns={'Charges_for_two':'cost','Ratings_out_of_5':'Ratings'})
zomato.columns
```

```python
#Some Transformations
zomato['cost'] = zomato['cost'].astype(str) #Changing cost to string 
zomato['cost'] = zomato['cost'].apply(lambda x: x.replace(',','.')) #Using lambda function to replace ',' from cost
zomato.info() #looking at the dataset information after transformation 
```

```python
#Reading unique values from the rating column
zomato['Ratings'].unique()
```

```python
#Checking the unique cost
zomato.cost.unique()
```

```python
#Encode the input variables
def Encode(zomato):
    for column in zomato.columns [~zomato.columns.isin(['Ratings','cost','Number of votes'])]:
        zomato[column].factorize()[0]
    return zomato

zomato_en = Encode(zomato.copy())
zomato_en.head(15)
```

```python
#Get correlation between different variables
corr = zomato_en.corr(method='kendall')
plt.figure(figsize=(14,8))
sns.heatmap(corr, annot=True)
zomato_en.columns
```

```python
#Defining the independent variables and dependent variables
x = zomato_en.loc[:,:]
y = zomato_en['Ratings']
#Getting test and training set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.1,random_state=353)
x_train.head()
```
