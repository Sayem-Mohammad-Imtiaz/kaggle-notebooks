```python
import markovify 
```

```python
rwsb = pd.read_csv("/kaggle/input/reddit-wallstreetsbets-posts/reddit_wsb.csv")
rwsb = rwsb[rwsb.body.notnull()]
rwsb = rwsb[rwsb.title.notnull()]
rwsb.head()
```

```python
rb = markovify.Text(rwsb['body'])
rt = markovify.NewlineText('\n'.join(rwsb['title']))
print("Done!")
```

```python
for x in range(10):
    print(rt.make_short_sentence(max_chars = 80))
    for x in range(2):
        print(rb.make_sentence())
    print(rb.make_sentence()+ "\n")
```
