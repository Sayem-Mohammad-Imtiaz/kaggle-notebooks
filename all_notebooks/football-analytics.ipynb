```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
xstat = pd.read_csv("../input/extended-football-stats-for-european-leagues-xg/understat.com.csv").rename(columns ={'Unnamed: 0': 'League','Unnamed: 1' : 'Year'})
xstat.info()
xstat.head()
```

```python
xstat1= xstat.query('team == "Liverpool"')
xstat1
```

```python
fig, ax = plt.subplots(figsize=(16, 8))

ax.barh(xstat1.Year, xstat1.ppda_coef, color='#189AB4', edgecolor='black')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_axisbelow(True)

plt.legend(['PPDA'])

plt.grid(alpha=0.3, axis='x')
plt.suptitle("Liverpool PPDA", ha='left', x='0.125', y=0.96, fontsize=18, fontweight='bold')
plt.show()
```

```python
xpts = xstat.pivot_table(index = "Year", columns = "team", values = "xpts_diff")
xpts[["Atletico Madrid", "Manchester City", "Sevilla", "Inter", "Tottenham"]].replace(0, np.nan).fillna(method = "bfill", axis = 0).plot(xlabel= "Year", ylabel = "Expected Goals Difference", title = "Expected Goals Difference For Top Teams By Year", figsize = (10,7))
```
