### This dataset(df1[ball by ball]) does not contain Season number. We will assign it

```python
conditions = [
    (df1['id']>=0) & (df1['id']< 336040),
    (df1['id']>392181) & (df1['id']< 392239),
    (df1['id']>=419106) & (df1['id']<= 419165),
    (df1['id']>=501198) & (df1['id']<= 501271),
    (df1['id']>= 548306) & (df1['id']<= 548381),
    (df1['id']>= 598006) & (df1['id']<= 598073),
    (df1['id']>=729279) & (df1['id']<=  734049),
    (df1['id']>= 829705) & (df1['id']<= 829823),
    (df1['id']>= 980901) & (df1['id']<= 981019), 
    (df1['id']>= 1082591)  & (df1['id']<= 1082650),
    (df1['id']>= 1136561)  & (df1['id']<= 1136620),
    (df1['id']>= 1175356)  & (df1['id']<= 1181768),
    (df1['id']>= 1216492)  & (df1['id']<= 1237181)
]

choices = [2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]

df1['season'] = np.select(conditions, choices, default = 2008)

```

```python
df1.sample(10)
```

NOW WE HAVE SEASON IN OUR DATASET DF1

### Season wise Orange Cap Holder

```python
df1.groupby(['season','batsman'])['total_runs'].sum().groupby('season').max().plot(kind = 'bar')
plt.rcParams["figure.figsize"] = (20,5)
plt.show()    

```
