```python
import plotly.graph_objects as go
```

```python
ar2 = pl.sort_values("Goal Per Shots On Target",ascending=False)["OnTarget"] #pl.sort_values("Goal Per Shots On Target",ascending=False)["Goals"]
```

```python
ar1 = pl.sort_values("Goal Per Shots On Target",ascending=False)["Goals"].values
```

```python
ar1 =  list(map(str, ar1))
ar2 =  list(map(str, ar2))
```

```python
ls = []
for i,a in zip(ar1,ar2):
    ls.append(i+"/"+a)
```

```python
pl = pl.sort_values("Goal Per Shots On Target",ascending=False)#pl.sort_values("Goal Per Shots On Target",ascending=False)
fig = go.Figure(data=[go.Bar(y=pl["Goal Per Shots On Target"], x=pl["Player Names"],
            hovertext=ls)])

fig.update_traces(marker_color='rgb(18,22,25)', marker_line_color='rgb(48,18,107)',
                  marker_line_width=1.5, opacity=0.6)
fig.update_layout(barmode='stack', xaxis={'categoryorder':'total descending'},yaxis_title="Accuracy",
                  title="Goals By Attempts On Target (Total Goals > 10)",
                  width = 1500
                 )


fig.show()
```

```python
sums = df.groupby("Player Names").sum()
```

```python
sums["M/G"] = sums["Mins"] / sums["Goals"]
sums = sums.reset_index()
sums = sums.sort_values("M/G")
min_goal = []
for a,i in zip(sums["Goals"],sums["Mins"]):
    min_goal.append(str(a) + "/" + str(i))
```

```python
#fig = px.histogram(sums.reset_index() , y="M/G", x = "Player Names")
#fig.update_layout(xaxis={"categoryorder":"total ascending"})

fig = go.Figure(data=[go.Bar(y=sums["M/G"], x=sums["Player Names"],
            hovertext=min_goal)])
fig.update_layout(barmode='stack', xaxis={'categoryorder':'total ascending'},yaxis_title="Goals",
                  title="Goals By Minutes",
                  width = 1500
                 )

fig.show()
```

```python
acc = df.groupby("Player Names").sum()
acc["Acc"] =  acc["OnTarget"] / acc["Shots"] 
acc = acc.sort_values("Acc",ascending=False)
```

```python
acc = acc[acc["Shots"]>15]
```

```python
ar1 = acc["OnTarget"].values
ar2 = acc["Shots"] 
```

```python
ar1 =  list(map(str, ar1))
ar2 =  list(map(str, ar2))
```

```python
ls = []
for i,a in zip(ar1,ar2):
    ls.append(i+"/"+a)
acc = acc.reset_index()
```

```python
fig = go.Figure(data = [go.Bar(y=acc["Acc"], x=acc["Player Names"],
            hovertext=ls)])
fig.update_layout(title="Most Accurate Shot Takers (Shots>15)")
fig.show()
```
