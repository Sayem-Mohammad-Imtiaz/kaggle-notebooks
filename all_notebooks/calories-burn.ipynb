```python
df=pd.read_csv("/kaggle/input/calories-burned-during-exercise-and-activities/exercise_dataset.csv",sep=",")
```

```python
df.head()
```

```python
df.columns=["Activity","lb_130","lb_155","lb_180","lb_205","calories_per_kg"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
sn.set_style(style="whitegrid")
```

```python
plt.subplots(1,1,figsize=(12,8))
sn.distplot(df.calories_per_kg,kde=True,color="c")
plt.show()
```

```python
sn.boxplot(df.calories_per_kg)
plt.show()
```

```python
sn.heatmap(df.corr(),annot=True,cmap="cool")
plt.show()
```

```python
best_results=df[(df.calories_per_kg > 3.0 ) & (df.calories_per_kg <= df.calories_per_kg.max())]
```

```python
serie_best_results=best_results.sort_values("calories_per_kg",ascending=False)
```

```python
serie_best_results
```

```python
plt.figure(figsize=(12,8))
plt.barh(serie_best_results.Activity,
        serie_best_results.calories_per_kg,
        color="c")

plt.show()
```
