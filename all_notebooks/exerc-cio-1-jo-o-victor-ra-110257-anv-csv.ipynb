# **Gráfico de barras analisando qual tipo de aeronave apresentou o maior número de ocorrências**

```python
quant_tipo = df.aeronave_tipo_veiculo.value_counts()

fig, ax = plt.subplots(figsize=(20,10))
ax.set_ylabel('Total de ocorrências', fontsize = '20')
ax.set_xticklabels(quant_tipo.index, fontsize = '20')
plt.xticks(rotation=45)

    
plt.tick_params(axis='x', which='major', labelsize=15)
plt.title('Total de Ocorrências por Tipo de Aeronave', fontweight='bold', color = 'orange', fontsize='30', horizontalalignment='center')

plt.bar(quant_tipo.index, quant_tipo.values)
```

# Análise de que, caso ocorra uma falha, qual manobra apresenta o maior número de fatalidades (mais perigosa/maior risco)

```python
fatal_fab = df.groupby('aeronave_fase_operacao').sum()
fatal_fab.head(3)
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.set_ylabel('Total de fatalidades', fontsize = '20')
ax.set_xticklabels(fatal_fab.index, fontsize = '20')
plt.xticks(rotation=90)
plt.tick_params(axis='x', which='major', labelsize=15)
plt.title('Total de Fatalidades por Manobra', fontweight='bold', color = 'orange', fontsize='30', horizontalalignment='center')
plt.grid()

plt.plot(fatal_fab.index, fatal_fab['total_fatalidades'])
```

## Com o gráfico acima podemos concluir que as fases de operação que apresentam o maior número de fatalidades caso ocorra uma falha são, em ordem decrescente: Cruzeiro, Decolagem e Manobra.

# Análise das características dos 30 modelos de aeronave que mais apresentaram falhas

```python
df.aeronave_motor_quantidade.unique()
```

## Mudando o valor da coluna 'aeronave_motor_quantidade' de **string** para **int**

```python
df.replace({'aeronave_motor_quantidade': {'MONOMOTOR': 1,
                                          'BIMOTOR': 2,
                                          'TRIMOTOR': 3,
                                          'SEM TRAÇÃO': 0,
                                          'QUADRIMOTOR': 4,
                                          '***': 0}},inplace = True)
```

## Obtendo os 30 modelos com maior número de ocorrências

```python
df.aeronave_modelo.value_counts().head(30).index
```

```python
lista_modelos = list(df.aeronave_modelo.value_counts().head(30).index)
df_model= df.loc[df['aeronave_modelo'].isin(lista_modelos)]
```

## Transformando a coluna 'aeronave_nivel_dano' de **string** para **int** seguindo o padrão:

* 0 - Nenhum dano ou indeterminado
* 1 - Dano leve
* 2 - Dano substancial
* 3 - Aeronave Destruída
 

```python
df_model.replace({'aeronave_nivel_dano': {'NENHUM': 0,
                                          'INDETERMINADO': 0,
                                          'LEVE': 1,
                                          'SUBSTANCIAL': 2,
                                          'DESTRUÍDA': 3}},inplace = True)
```

```python
import plotly.express as px

fig = px.parallel_coordinates(df_model, color="aeronave_nivel_dano",
                              dimensions=['aeronave_motor_quantidade', 
                                          'aeronave_assentos', 
                                          'aeronave_pmd', 
                                          'aeronave_ano_fabricacao'],
                              range_color = [0,3],
                              color_continuous_scale=px.colors.diverging.Tealrose,
                              color_continuous_midpoint=2)
fig.show()
```

## Assim podemos notar que as aeronaves que apresentaram ocorrências mais graves, sendo destruídas, são majoritariamente de porte pequeno, monomotores, com menos de 20 assentos e fabricadas após os anos 2000.

# Análise de qual páis e fabricante são as aeronaves com maior número de ocorrências

```python
df.aeronave_fabricante = df.aeronave_fabricante.astype(str)
count = df.aeronave_fabricante.value_counts()
```

## Criação de um Wordcloud para verificar quais fabricantes apresentaram mais ocorrências

```python
import wordcloud as w

d = dict(zip(count.index, count.values))

wordcloud = w.WordCloud(width=2000, height=800,
                        margin=0, 
                        collocations=False, 
                        background_color="black",
                        stopwords = 'nan').generate_from_frequencies(d)
```

```python

plt.figure(figsize=(15,8))
plt.imshow(wordcloud, interpolation='bilinear', aspect = 'equal')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

## Com o wordcloud podemos ter uma noção das fabricantes que apresentaram maiores ocorrências, para confirmar vamos plotar um gráfico de barras 

```python
fig, ax = plt.subplots(figsize=(20,10))
ax.set_xticklabels(count.index[:10], fontsize = '20')
plt.xticks(rotation=45)

    
plt.tick_params(axis='x', which='major', labelsize=15)
plt.title('Total de Ocorrências por Fabricante', fontweight='bold', color = 'orange', fontsize='30', horizontalalignment='center')

plt.bar(count.index[:10], count.values[:10])
```

# Pie plot apresentando os países de origem das aeronaves que mais tiveram ocorrências

```python
lista_paises = list(df.aeronave_pais_fabricante.value_counts().head(10).index)
df_paises = df.loc[df['aeronave_pais_fabricante'].isin(lista_paises)]

count = df_paises.aeronave_pais_fabricante.value_counts()
g1 = list(count.index)
v1 = list(count.values)
```

```python

fig, ax = plt.subplots()
ax.axis('equal')

mypie, _ = ax.pie(v1, labels = v1,radius=2 ,wedgeprops = { 'linewidth' : 5, 'edgecolor' : 'white' })

my_circle=plt.Circle( (0,0), 1.1, color='white')
p=plt.gcf()
p.gca().add_artist(my_circle)

ax.legend(mypie, g1,
          title="Países",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

ax.set_title('Ocorrências por país \nde origem da fabricante', loc='center', y = 0.42)

plt.figure(figsize=(20,15))
plt.show()
```
