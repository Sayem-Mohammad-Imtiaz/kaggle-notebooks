```python
X_train = np.array(train_data[train_data.columns[2:-1]])
y_train = np.array(train_data[train_data.columns[-1]])

X_test = np.array(test_data[test_data.columns[2:-1]])
y_test = np.array(test_data[test_data.columns[-1]])
```

```python
y_test
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
Correctness = accuracy_score(y_test, y_pred)
print(f'Your model is {Correctness*100}% accurate!')
```

```python
sns.countplot(y_pred,saturation=10)
plt.rc('figure',figsize=(19,12))
```

Using decision tree

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
```

```python
dt_ypred = dt_model.predict(X_test)
dt_pred_score = accuracy_score(y_test, dt_ypred)
print(f'The Decision Tree accuray is {dt_pred_score*100}%')
```

Using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier()

rf_model.fit(X_train, y_train)
```

```python
rf_ypred = rf_model.predict(X_test)
```

```python
rf_accuracy = accuracy_score(y_test, rf_ypred)
print(f'Random Forest accuracy is {rf_accuracy*100}%')
```
