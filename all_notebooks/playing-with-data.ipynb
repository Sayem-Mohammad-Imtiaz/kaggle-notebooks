Let's import our data

```python
def conv_str_to_datetime(x):
    return(datetime.strptime(x, '%Y%m%d %H:%M:%S.%f'))

df = pd.read_csv('/kaggle/input/gbpusd-price-and-tick-volume/2021.6.16GBPUSD-TICK-NoSession.csv', header=0,names=['DateTime', 'Bid', 'Ask', 'Volume'])
```

Now we need out dataframe to be the DateTime column

```python
df['DateTime']= pd.to_datetime(df['DateTime'],format='%Y%m%d %H:%M:%S.%f',errors="coerce")
df = df.set_index("DateTime")
```

We have tick data but we can resample to higher time frames.
Let's resample to 15 minutes timeframe.

```python
ask = df['Ask'].resample('15Min').ohlc()
```

```python
ask.head
```

Finally let's plot our resampled data 

```python
mpf.plot(ask[-100:-1], type='candle')
```
