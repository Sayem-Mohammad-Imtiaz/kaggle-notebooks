# import necessary libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df = pd.read_csv("/kaggle/input/students-performance-in-exams/StudentsPerformance.csv") # getting data
df
```

# looking data

```python
df.info()
```

```python
plt.figure(figsize = (16,8))
sns.pairplot(df,hue  = "gender")
plt.show()
```

```python
grp = df.groupby("gender").aggregate(["mean"]) # group the gender
grp
```

woman are better at reading and writing than men.
men are better at math than woman

# visualization of gender

```python
plt.figure(figsize = (14,6))
plt.subplot(121)
sns.barplot(grp.index,grp['writing score','mean'])
plt.subplot(122)
sns.barplot(grp.index,grp['reading score', 'mean'])
plt.xlabel("Gender")
plt.show()
plt.figure(figsize = (14,7))
labels = ["Female","Male"]
plt.pie(grp['math score','mean'], autopct="%.1f%%", labels=labels, pctdistance=0.5)
plt.title("Math mean score")
plt.show()
```

male's math > female's math <br>
female's reading > male's reading <br>
female's writing  > male's writing

```python
plt.figure(figsize = (14,6))
plt.subplot(121)
sns.distplot(df['math score'])
plt.subplot(122)
sns.distplot(df['reading score'])
plt.show()
plt.figure(figsize = (14,6))
sns.distplot(df['writing score'])
plt.show()
```

```python
lsn = df.groupby(["parental level of education"]).aggregate(["mean"]) # group the level of  education
lsn
```

# visualization the level of education

```python
plt.figure(figsize = (14,6))
labels = [i for i in lsn.index]
plt.pie(lsn['math score','mean'], autopct="%.1f%%",explode=[0.05]*6, labels=labels, pctdistance=0.5)
plt.title("Math mean score")
plt.show()
```

most succesfull level of education => bachlor's degree and master's degree <br>
worst level of education => high school

```python
plt.figure(figsize = (16,6))
plt.subplot(121)
sns.barplot(lsn.index,lsn['writing score','mean'])
plt.xticks(rotation = 60)
plt.subplot(122)
sns.barplot(lsn.index,lsn['reading score','mean'])
plt.xticks(rotation = 60)
plt.show();
```

```python
plt.figure(figsize = (16,7))
sns.barplot(x = "parental level of education", y = "math score", data = df, hue = "gender")
plt.show()
```

male always more succesfull math than female

```python
for i in range(len(df["test preparation course"])):
    if df.loc[i,"test preparation course"] == "none":
        df.loc[i,"test preparation course"] = "not completed"
df
```

```python
plt.figure(figsize = (16,7))
sns.barplot(x = "test preparation course", y = "math score", data = df,hue = "gender")
plt.show()
```
