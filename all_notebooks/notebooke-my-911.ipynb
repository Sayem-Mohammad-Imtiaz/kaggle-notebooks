```python
df['timeStamp']=pd.to_datetime(df['timeStamp'])
type(df['timeStamp'].iloc[0])
```

```python
time = df['timeStamp'].iloc[0]
print(time.hour)
print(time.dayofweek)
print(time.year)
```

```python
df['DayOfWeek'] = df['timeStamp'].apply(lambda time:time.dayofweek)
```

```python
df.head()
```

```python
df['DayOfWeek'].unique()
```

```python
dic_dayOfWeek= {0:'mon',1:'tue',2:'wed',3:'thu',4:'fri',5:'sat',6:'sun'}
```

```python
df['DayOfWeek'] = df['DayOfWeek'].map(dic_dayOfWeek)
```

```python
df.head()
```

```python
sns.countplot(data=df,x='DayOfWeek',hue='reason')
```
