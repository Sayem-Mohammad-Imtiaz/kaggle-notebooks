```python
data = pd.DataFrame(data, columns = ['Province/State', 'Region code', 'Country/Region', 'Lat', 'Long'])
data.hist()
plt.show()
```

```python
data.plot.bar() 

# plot between 2 attributes 
plt.bar(data['Lat'], data['Long']) 
plt.xlabel("Lat") 
plt.ylabel("Long") 
plt.show() 
```
