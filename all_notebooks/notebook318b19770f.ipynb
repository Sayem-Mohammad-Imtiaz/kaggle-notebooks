```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Factors that affect happiness around the globe :

```python
df = pd.read_csv('../input/world-happiness-report-2021/world-happiness-report.csv')
```

```python
df.head()
```

```python
df.corr()['Positive affect']
```

```python
plt.figure(figsize = (12,8),dpi = 200)
sns.scatterplot(x='Freedom to make life choices', y='Positive affect',data= df)
```

```python
plt.figure(figsize = (12,8),dpi = 200)
sns.scatterplot(x='Freedom to make life choices', y='Positive affect',data= df,hue = 'Perceptions of corruption')
```

```python
df.corr()['Negative affect']
```

Here as we can see that hapiness is most positively correlated with frredom of making choices i.e as freedom of making choices increase happiness increases


And we also observe that people have the most negative affect on happiness with geneorisity and also people have least negative affect with social support.

```python
plt.figure(figsize = (12,8), dpi = 200)
sns.scatterplot(x='Social support', y = 'Negative affect', hue = 'Generosity',data = df)
```

 **As we can observe that with the increase in social support negative affect decreases**


# Now, lets explore the factors affecting happiness in India

```python
India_df = df[(df['Country name'] == 'India')]
```

```python
India_df.head()
```

```python
India_df.isnull().sum()
```

```python
India_df['Generosity'].isnull()
```

**here we can see that at index = 732 , generosity value is not available we can ignore this for now , as we are not dealing with feature engineering**

```python
India_df.corr()['Positive affect']
```

**we can see that which factor most positively correlates with the Positive affect of Happiness in India**

```python
plt.figure(figsize = (12,8),dpi =200)
sns.scatterplot(x='Social support', y= 'Positive affect', data = India_df,hue ='Generosity')
```

```python
India_df.corr()['Negative affect']
```

**People in India have most negative affect on happiness with healthy life expectancy at birth**

**And have least negative affect on happiness with life ladder**

```python
plt.figure(figsize =(12,8), dpi = 200)
sns.scatterplot(x= 'Healthy life expectancy at birth',y='Negative affect',data = India_df, hue='Life Ladder')
```
