```python
# replacing income column with 0 and 1
data1['income']=data1['income'].replace((' <=50K',' >50K'),(0,1))
```

```python
data1
```

```python
# removing duplicate rows 
data1.drop_duplicates(inplace=True)
data1
```

```python
# droping column 'education' since education-num column convay the same.
data1.drop(['education'],axis=1,inplace=True)
```

```python
# doing one hot encoding to change categorical variables to numerical variables
data2=pd.get_dummies(data1)
data2
```

```python
data2.columns
```

```python
# droping target variable
data2.drop(['income'],axis=1,inplace=True)
```

```python
# separating dependent and independent variables
x=data2.values
x
```

```python
y=data1.loc[:,'income']
y
```

```python
# splitting data into train set and test set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=.25)
```

```python
# standardising the data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
# applying models
```

```python
# Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
lrc=LogisticRegression(random_state=0)
lrc.fit(x_train,y_train)
y_pred1=lrc.predict(x_test)
y_pred1
```

```python
from sklearn.metrics import * 
```

```python
# confusion matrix
```

```python
confusion_matrix(y_test,y_pred1)
```

```python
print(classification_report(y_test,y_pred1))
```

```python
y_problog=lrc.predict_proba(x_test)[:,1]
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=40,random_state=0)
rfc.fit(x_train,y_train)
y_pred2=rfc.predict(x_test)
y_pred2
```

```python
# calculate the fpr and tpr for all thresholds of the classification
# Create ROC Graph
import sklearn.metrics as metrics
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, y_problog)
logit_roc_auc=metrics.auc(fpr,tpr)
rf_fpr, rf_tpr, rf_thresholds = roc_curve(y_test, y_probrfc)
rf_opt_roc_auc=metrics.auc(rf_fpr,rf_tpr)
gb_fpr, gb_tpr, gb_thresholds = roc_curve(y_test, y_probgbc[:,1])
gb_opt_roc_auc=metrics.auc(gb_fpr,gb_tpr)
plt.figure(figsize=(14, 6))

# Plot Logistic Regression ROC
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
# Plot Random Forest ROC
plt.plot(rf_fpr, rf_tpr, label='Random Forest (area = %0.2f)' % rf_opt_roc_auc)
# Plot Gradient Boosting ROC
plt.plot(gb_fpr, gb_tpr, label='Gradient Boosting(area = %0.2f)' % gb_opt_roc_auc)

# Plot Base Rate ROC
plt.plot([0,1], [0,1],label='Base Rate' 'k--')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Graph')
plt.legend(loc="lower right")
plt.show()
```
