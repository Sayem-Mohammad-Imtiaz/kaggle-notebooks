```python
import pandas as pd
from AutoTS.AutoTS import AutoTS
```

```python
df = pd.read_csv(r"../input/gold-prices/annual_csv.csv",parse_dates=["Date"])
```

```python
df.index = df["Date"] 
df
```

```python
model = AutoTS()
```

```python
model.fit(df,series_column_name="Price")
```

```python
model.predict(start=pd.to_datetime('2020-12-01'), end=pd.to_datetime('2050-12-01'))
```
