```python
df_housing.drop(['Address'],axis = 1,inplace = True)
```

```python
df_housing.head()
```

```python
train_count = int(df_housing.shape[0] * 0.7)
test_count = df_housing.shape[0] - train_count
train_count,test_count
```

```python
Y_test = df_housing.iloc[:test_count,5]
Y_train = df_housing.iloc[test_count:,5]
Y_train.shape, Y_test.shape
```

```python
Y_test.head()
```

```python
X_train = df_housing.iloc[:train_count,[0,1,2,3,4]]
X_test = df_housing.iloc[train_count:,[0,1,2,3,4]]
X_train.shape, X_test.shape
```

```python
X_test.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression().fit(X_train, Y_train)
```

```python
print("Coefficient: ",reg.coef_)
print("Intercept: ",reg.intercept_)
```

```python
Y_pred = reg.predict(X_test)
```

```python
rmse = (sum((Y_pred - Y_test)**2)/len(Y_test))**0.5
print("Root Mean Squared Value: ", rmse)
```
