It is a multi-class classification problem to predict which product the complaint is about

```python
train = pd.read_csv("/kaggle/input/us-consumer-finance-complaints/consumer_complaints.csv")
train.head()
```

```python
train.shape
```

```python
train = train[train.consumer_complaint_narrative.notnull()]
```

```python
train.shape
```

```python
train['len_text'] = train.consumer_complaint_narrative.apply(lambda x: len(x.split()))
```

```python
train = train[train.len_text > 249]
```

```python
train.shape
```

```python
train = train[['consumer_complaint_narrative', 'product']]
train.reset_index(inplace = True, drop = True)
train.head()
```

```python
train = train.rename(columns = {'consumer_complaint_narrative' : 'text', 'product': 'label'})
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
train['label'] = LE.fit_transform(train['label'])
train.head()
```

```python
import re
def clean_txt(text):
    text = re.sub("'","",text)
    text = re.sub("(//W)+"," ",text)
    return text 
```

```python
train['text'] = train.text.apply(clean_txt)
train.head()
```

Formatting the data for BERT model

```python
def get_split(text1):
    l_total = []
    l_parcial = []
    if len(text1.split())//150 >0:
        n = len(text1.split())//150
    else: 
        n = 1
    for w in range(n):
        if w == 0:
            l_parcial = text1.split()[:200]
            l_total.append(" ".join(l_parcial))
        else:
            l_parcial = text1.split()[w*150:w*150 + 200]
            l_total.append(" ".join(l_parcial))
    return l_total
```

```python
train['text_split'] = train['text'].apply(get_split)
```
