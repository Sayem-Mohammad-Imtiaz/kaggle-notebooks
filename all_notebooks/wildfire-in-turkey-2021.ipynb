**MEGER DATA**

```python
all_loc = [df1,df2,df3,df4,df5,df6,df7,df8]
loc_merge = pd.concat(all_loc, ignore_index=True)
loc_merge.info()
loc_merge
```

```python
#fillter just vegetation_fire
loc_merge['acq_date'] = pd.to_datetime(loc_merge['acq_date'])
vegetation_fire= loc_merge [loc_merge["type"] == 0]
fire_2021 = vegetation_fire[vegetation_fire['acq_date']>'2021-01-01']
fire_2021.head()
```

#### LOCAL LATITUDE AND LONGITUDE

### LOCATIONS

```python
import folium
from folium import plugins

heat = list(zip(fire_2021.latitude,fire_2021.longitude))
MAP_FUNCTION = folium.Map(location=[38.9637,35.2433],
                  tiles='Stamen Terrain',
                  zoom_start=4.5)

plugins.MarkerCluster(heat).add_to(MAP_FUNCTION)

MAP_FUNCTION
```

#### BRIGHTNESS

```python
wildfire=fire_2021.groupby(['acq_date','latitude','longitude']).mean()
wildfire
```

```python
wildfire[['brightness']].plot(xlabel='date and location',
                              ylabel='brightness temperature (kelvin)')
plt.xticks(rotation=85)
plt.show()
```

```python
plt.figure(figsize=(100,500))
wildfire[['bright_t31']].plot(xlabel='date and location',
                              ylabel='brightness temperature (kelvin)')
plt.xticks(rotation=85)
plt.show()
```
