```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from pandas_profiling import ProfileReport
```

```python
data = pd.read_csv('../input/covid19-cases-in-214-countries/covid-19  cases around the world.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['date']=pd.to_datetime(data['date'])
```

```python
plt.figure(figsize=(20,7))
plt.style.use("dark_background")
sns.lineplot(x='date',y='new_cases',data=data)
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(x='date',y='total_cases',data=data)
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(x='date',y='total_deaths',data=data)
```

```python
data.groupby("continent").total_cases.sum().plot(figsize=(12,7))
```

```python
df = data[(data["continent"]=="North America") |(data["continent"]=="Asia")|(data["continent"]=='Africa') |(data["continent"]=='Europe')|(data["continent"]=='South America')|(data["continent"]=='Oceania')]
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=df,x="date",y="total_cases",marker='o',hue="continent")
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=df,x="date",y="new_cases",marker='o',hue="continent")
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=df,x="date",y="total_deaths",marker='o',hue="continent")
plt.show()
```
