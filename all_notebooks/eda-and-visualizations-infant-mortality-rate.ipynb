```python
df = pd.read_csv("../input/infant-mortality/InfantMortalityRate.csv")
df.head()
```

```python
df.shape
```

```python
df.Gender.value_counts()
```

```python
df['Gender'] = df['Gender'].map({'Female': 0, 'Male': 1, 'Total': 2})
```

```python
df.Gender.value_counts()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.Country.value_counts()[:20]
```

```python
df[['Infant Mortality Rate', 'Gender', 'Country','Year']].sort_values('Infant Mortality Rate', ascending = False)[:30]
```

## Highest infant mortality rate is in country Sierra Leona and the year is 2009
## Highest infant mortality rate for Male is 116.93 in year 2009
## Highest infant mortality rate for Female is 104.33 in year 2009

```python
df[['Infant Mortality Rate', 'Gender', 'Country','Year']].sort_values('Infant Mortality Rate', ascending = False)[-30:]
```

## Lowest infant mortality rate is in country San Marino and the year is 2019
## Lowest infant mortality rate for Female is 1.36 in San Marino in year 2019
## Lowest infant mortality rate for Male is 1.61 in San Marino in year 2019

```python
df[df['Year']==2019.0][['Infant Mortality Rate', 'Country', 'Gender']].sort_values('Infant Mortality Rate',ascending = False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams.update({'figure.figsize': (10, 7), 'figure.dpi': 120})
```

```python
sns.barplot(y = df['Infant Mortality Rate'], x =df['Gender'], data=df)
```

### Infant Mortality Rate of Male is higher

```python
sns.barplot(y = df['Infant Mortality Rate'], x =df['Year'], data=df)
```

```python
IMR100 = df[df['Infant Mortality Rate']>100]
```

```python
IMR100
```

```python
sns.barplot(y = IMR100['Infant Mortality Rate'], x =IMR100['Country'], hue = IMR100['Year'], data=df)
```

### Countries with Mortality Rate above 100 are Sierra Leone, Central African Republic, Somalia

```python
sns.catplot(y = IMR100['Infant Mortality Rate'], x =IMR100['Gender'], data=df)
```

```python
contrUS = df[df['Country']=='United States']
contrUS
```

```python
sns.barplot(y = contrUS['Infant Mortality Rate'], x =contrUS['Country'],hue = contrUS['Year'], data=df)
```

```python
contrind = df[df['Country']=='India']
contrind
```

```python
sns.barplot(y = contrind['Infant Mortality Rate'], x =contrind['Country'],hue = contrind['Year'], data=df)
```
