# Exploring Null and Unique values distribution

```python
def NullUnique(df):
    dic = defaultdict(list)
    for col in df.columns:
        dic['Feature'].append(col)
        dic['NumUnique'].append(len(df[col].unique()))
        dic['NumNull'].append(df[col].isnull().sum())
        dic['%Null'].append(round(df[col].isnull().sum()/df.shape[0] * 100,2))
    return pd.DataFrame(dict(dic)).sort_values(['%Null'],ascending=False).style.background_gradient()
```

```python
NullUnique(data)
```

# Is the industry really Sexist?

```python
fem_sal_norm = data.groupby(['gender']).agg('sum')['salary'].F/data.loc[(data.gender=="F") & (data.status=="Placed") & (data.salary>0)].shape[0]
male_sal_norm = data.groupby(['gender']).agg('sum')['salary'].M/data.loc[(data.gender=="M") & (data.status=="Placed") & (data.salary>0)].shape[0]
```

```python
fig = go.Figure([go.Bar(x=['Male Normal Salary','Female Normal Salary'], y=[male_sal_norm,fem_sal_norm])])
fig.update_layout(title='Normal Salaries for Men vs Women')
fig.show()
```

**Observations**:
- The Mean salary for men is 31k more than women
- Indicating Men in general are paid more than women

# Does 10th Std Board make a Difference to Placements?

```python
fig = go.Figure(data=[
    go.Bar(name='Not Placed', x=['Central','Others'], y=[data.groupby(['ssc_b']).status.value_counts().Central['Not Placed'] , data.groupby(['ssc_b']).status.value_counts().Others['Not Placed']]),
    go.Bar(name='Placed', x=['Central','Others'], y=[data.groupby(['ssc_b']).status.value_counts().Central['Placed'] , data.groupby(['ssc_b']).status.value_counts().Others['Placed']])
])
fig.update_layout(barmode='group',title='Does SSC Board make a difference?')
fig.show()
```

```python
print(f"Ratio of Placed to Not Placed\nCentral\t{data.groupby(['ssc_b']).status.value_counts().Central['Placed']/data.groupby(['ssc_b']).status.value_counts().Central['Not Placed']}\nOthers\t{data.groupby(['ssc_b']).status.value_counts().Others['Placed']/data.groupby(['ssc_b']).status.value_counts().Others['Not Placed']}")
```

**Observations**:
- There is no significant difference however students from other boards have a higher Place to Not Placed ratio indicating slight advantage

# Does SSC Percentage Influence Salary and Placements?

```python
fig = px.box(data, x="status", y="ssc_p")
fig.update_layout(title='Does SSC % influence Placed or Not')
fig.show()
```

**Observations**:
- SSC% clearly makes a difference, the median % for Placed students is 72.5% whereas for Not Placed it is 56.28%
- For Placed students % lies in : 49 - 89% range
- For Not Placed students % lies in : 40 - 77.8% range


```python
 sns.jointplot("ssc_p", "salary", data=data, kind="reg")
```

**Observations**:
- Higher salaries were bagged by students around and above the median %
- Students with higher ssc_p are more likely to get paid more than lower scorers as shown by the regression line
