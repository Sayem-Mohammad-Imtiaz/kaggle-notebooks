```python
#Apurva Misra-20788126
#Based on dataset from https://www.nature.com/articles/s41597-020-0448-0
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
from wordcloud import WordCloud
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/beoutbreakprepared/nCoV2019/master/latest_data/latestdata.csv')
df.head(5)
```

```python
df.columns=df.columns.str.replace(" ","")
df.columns=[x.lower() for x in df.columns]
```

```python
df.columns
```

```python
df.drop(["id"],axis=1,inplace=True)
```

```python
len(df)
```

```python
df.dropna(axis=0,how='all',inplace=True)
```

```python
len(df)
```

```python
df.info()
```

```python
#processing ages
age_num=[]
for age_str in df.age:
    if "-" in str(age_str):
        a,b=str.split(age_str,"-")
        if(b==""):
            b=0
        age_num.append((int(a)+int(b))/2)
    else:
        age_num.append(age_str)
df.age=age_num
```

```python
df_gender=df[df.sex.isna()!=True]
```

# People affected in different countries differentiated by gender

```python
fig=plt.figure(figsize=(22, 16), dpi= 80, facecolor='w', edgecolor='k')
ax = sns.countplot(x="country", hue="sex", data=df_gender)
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
plt.show()
```

```python
df.age.unique()
df.age.replace('6 weeks', 0.125, inplace=True)
df.age.replace('13 month', 1.08, inplace=True)
df.age.replace('18 month', 1.5, inplace=True)
df.age.replace('5 months', 0.416, inplace=True)
df.age.replace('5 month', 0.416, inplace=True)
df.age.replace('8 month', 0.66, inplace=True)
df.age.replace('6 months', 0.5, inplace=True)
df.age.replace('9 month', 0.75, inplace=True)
```

```python
reduced=df[df.age.isna()!=True]
reduced.age=pd.to_numeric(reduced.age)
#print(len(reduced))
#print(reduced.outcome.unique())
#print(len(reduced[reduced.outcome.isna()!=True]))
```

```python
reduced=reduced[reduced['outcome'].isin(['death','stable','hospitalized'])]
```

# Outcome of affected people under 50 

```python
print(reduced["outcome"].where((reduced.age<50)).value_counts())
ax = sns.countplot(x="outcome", data=reduced.where((reduced.age<50)))
plt.xticks(rotation=90)
```

# Outcome of affected people between 50 and 75

```python
print(reduced["outcome"].where((reduced.age>50)).value_counts())
ax = sns.countplot(x="outcome", data=reduced.where((reduced.age>50) & (reduced.age<75)))
plt.xticks(rotation=90)
```

# Outcome of affected people above 75

```python
print(reduced["outcome"].where((reduced.age>50)).value_counts())
ax = sns.countplot(x="outcome", data=reduced.where((reduced.age>75)),order=["stable","death","hospitalized"])
plt.xticks(rotation=90)
```
