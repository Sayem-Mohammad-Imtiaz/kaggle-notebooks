# Analysing the relationship of Height and Weight with Gender

## Import packages

## Import data

```python
##Importing the packages
#Data processing packages
import numpy as np 
import pandas as pd 

#Visualization packages
import matplotlib.pyplot as plt 
import seaborn as sns 

#Suppress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("../input/height-weight-gender/heights_weights.csv")
data.head()
```

**Display Information (info()) about the dataset**

```python
data.info()
```

**Display size (shape) about the dataset**

```python
data.shape
```

**Visualize the 'data' using 'pairplot' with hue as "Gender", height=3 and kind='scatter'**

```python
sns.pairplot(data=data,hue='Gender',height=3, kind='scatter')
```

<font color='blue'>**OBSERVATION :** There is strong correlation between height and gender</font>

**Visualize the 'data' using 'boxplot' - 'Gender' vs 'Height'**

```python
sns.boxplot(x='Gender', y='Height', data=data)
```

<font color='blue'>**OBSERVATION :** The height of males is more than females</font>

**Visualize the 'data' using 'boxplot' - 'Gender' vs 'Weight'**

```python
sns.boxplot(x='Gender', y='Weight', data=data)
```

<font color='blue'>**OBSERVATION :** The weight of males is more than females</font>
