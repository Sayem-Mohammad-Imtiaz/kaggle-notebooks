```python
# Loading Essentail Libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sb
import plotly
from plotly.offline import plot, iplot, download_plotlyjs, init_notebook_mode
init_notebook_mode(connected=True)
import re

# loading data
dataFrameYearWise = pd.read_csv('/kaggle/input/world-population/world_population_by_year.csv')
dataFrameCountryWise = pd.read_csv('/kaggle/input/world-population/world_population_by_countries.csv')
dataFrameRegionWise = pd.read_csv('/kaggle/input/world-population/world_population_by_region.csv')

# interpolating nan values
# sorting in acsending order
dataFrameYearWise = dataFrameYearWise.fillna(method = 'ffill')
dataFrameYearWise.sort_values(by = 'Year (July 1) ', inplace = True)

#conveting string to intger vals
dataFrameCountryWise['Population'] = dataFrameCountryWise['Population'].map(lambda a: re.sub(',','',a)).astype('int')
```

```python
# Procceding with task one
# lets define a function for plotting 
def ParamPlotter(dataFrame, paramName, docString):
    years = np.array (dataFrame['Year (July 1) '])
    param = np.array(dataFrame[paramName])
    
    plt.figure(figsize = (15, 6))
    plt.plot(years, param, color = 'green')
    plt.scatter(years, param, color = 'blue')
    plt.title(docString)
    plt.grid(True)
    plt.show()
```

```python
ParamPlotter(dataFrameYearWise, 'Population', ' Fig1: variation of World Population 1955-2020')
```

```python
ParamPlotter(dataFrameYearWise, 'Yearly %  Change', ' Fig:2 %  change in world Pop 1955-2020')
```

```python
ParamPlotter(dataFrameYearWise, 'Fertility Rate', ' Fig:3 Variation of World Fertility Rates 1955-2020')
```

```python
ParamPlotter(dataFrameYearWise, 'Median Age', ' Fig:4 Variation of World median age  1955-2020')
```

```python
ParamPlotter(dataFrameYearWise, 'Density (P/KmÂ²)', ' Fig:5 Variation of World\'s Population Density P/km^2 1955-2020')
```

```python
ParamPlotter(dataFrameYearWise, 'Urban Pop %', ' Fig:6 Variation of World\'s  Urban Pupulation % wise 1955-2020')
```

# Diving a level deeper: World Population analysis CountryWise
lets use plotly's chloropeth maps for this Purpose

```python
import plotly.express as px

# getting country names and iso values
# needed for plotting chloropleth maps
# Note this is only one way of doing this
# other might exists.
df = px.data.gapminder().query("year==2007")[['country', 'iso_alpha']].values
mapper = {}
for mem in df:
    mapper[mem[0]] = mem[1]

mapper['Russia'] = 'RUS'
dataFrameCountryWise['iso_values'] = dataFrameCountryWise['Country']
dataFrameCountryWise['iso_values'] = dataFrameCountryWise['iso_values'].map(mapper)

# Now Plotting chloropleth map 
# starting with World Population Countriwise
fig = px.choropleth(dataFrameCountryWise, locations="iso_values",
                    color="Population", 
                    hover_name="Country",
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```

We can observe China and India have the largest Population[](http://)

```python
# lets now Plot median age
fig = px.choropleth(dataFrameCountryWise, locations="iso_values",
                    color="Med. Age", 
                    hover_name="Country",
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```

Talking about median Ages:
* India has median Population age of 28.
* USA, China, Russia, Canada have median population ages : 38, 38, 40, 41 years respectively

Lets now talk about fertility Rates

```python
fig = px.choropleth(dataFrameCountryWise, locations="iso_values",
                    color="Fert. Rate", 
                    hover_name="Country",
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```

This map just indicates the fact: Developing Nations have higher Fertility Rates than Developed Nations, as the countries progress further it is excepted that their fertility rates will drop as well (I guess fertility rate of around 2 will do the trick.)

Now Taking Migration of People into consideration  into consideration

```python
fig = px.choropleth(dataFrameCountryWise, locations="iso_values",
                    color="Migrants (net)", 
                    hover_name="Country",
                    color_continuous_scale=px.colors.sequential.Plasma)
fig.show()
```
