```python
import numpy as np 
import pandas as pd
from sklearn import preprocessing
```

```python
raw_csv_data = np.loadtxt('/kaggle/input/audiobook-sales-prediction/Audiobooks_data.csv', delimiter = ',')
```

**Extract the data from csv**

```python
unscaled_input_all = raw_csv_data[:,1:-1]
target_all = raw_csv_data[:,-1]
```

**balance the dataset**

```python
num_one_targets = int(np.sum(target_all))
zero_target_counter = 0
indices_to_remove = []

for i in range(target_all.shape[0]):
    if target_all[i] == 0:
        zero_target_counter += 1
        if zero_target_counter > num_one_targets:
            indices_to_remove.append(i)
unscaled_input_equal_prior = np.delete(unscaled_input_all, indices_to_remove, axis = 0)
targets_equal_prior = np.delete(target_all, indices_to_remove, axis = 0)
            
```

**standardize the input**

```python
scaled_input = preprocessing.scale(unscaled_input_equal_prior)
```

**shuffle the data**

```python
shuffled_indices = np.arange(scaled_input.shape[0])
np.random.shuffle(shuffled_indices)
shuffled_inputs = scaled_input[shuffled_indices]
shuffled_targets = targets_equal_prior[shuffled_indices]
```
