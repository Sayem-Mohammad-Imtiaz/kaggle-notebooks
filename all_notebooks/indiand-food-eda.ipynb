# 4.1 Diet

```python
#Checking if everything is fine with diet column
df['diet'].value_counts()
```

```python
#This code was done similarly as Joshua did

#Counting ratio
x_diet = df.groupby('diet')['diet'].count()
y = len(df)
ratio = (x_diet/y).round(2)

diet_ratio = pd.DataFrame(ratio).T

#Plotting
fig, ax = plt.subplots(1,1,figsize=(10, 5))

ax.barh(diet_ratio.index, diet_ratio['vegetarian'], color = '#aee378')
ax.barh(diet_ratio.index, diet_ratio['non vegetarian'], left=diet_ratio['vegetarian'], color = '#b3646f')

#Modifying axis
ax.set_xlim(0, 1)
ax.set_xticks([])
ax.set_yticks([])

#Annotating
#For vegetarian
for i in diet_ratio.index:
    ax.annotate(f"{int(diet_ratio['vegetarian'][i]*100)}%",
                xy=(diet_ratio['vegetarian'][i]/2,i),
               va = 'center', ha ='center', fontfamily = 'serif', fontsize = 30, color = 'black')
    
    ax.annotate("Vegetarian", xy = (diet_ratio['vegetarian'][i]/2, -0.15),
               va = 'center', ha = 'center', fontfamily = 'serif', fontsize = 15)

#For non vegetarian   
for i in diet_ratio.index:
    ax.annotate(f"{int(diet_ratio['non vegetarian'][i]*100)}%",
               xy = (diet_ratio['vegetarian']+diet_ratio['non vegetarian'][i]/2, i), fontsize = 15, va = 'center',
               ha = 'center', fontfamily = 'serif')
    
    ax.annotate('Non vegetarian', xy = (diet_ratio['vegetarian'] + diet_ratio['non vegetarian'][i]/2, - 0.15),
               va = 'center', ha = 'center', fontfamily = 'serif', fontsize = 10)

#Adding text
fig.text(0.125, 1.03, 'Vegeterian and non vegeterian dishes ratio', fontfamily = 'serif', fontsize = 30, fontweight = 'bold')
fig.text(0.125, 0.95, 'We see that most of the indian dishes are vegetarian', fontfamily = 'serif', fontsize = 20)

#Removing spines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)

plt.show()
```

# 4.2 Preparation time

```python
fig, ax= plt.subplots(1,1,figsize=(20, 5))

#Adding distribution graph and mean line for vegetarian dishes
sns.distplot(df[df['diet'] == 'vegetarian']['prep_time'], kde = True, kde_kws = {'linewidth': 2}, color = '#aee378', label = 'Vegetarian')
ax.axvline(df[df['diet']=='vegetarian']['prep_time'].mean(), color = 'green')

#Adding distribution graph and mean line for non vegetarian dishes
sns.distplot(df[df['diet']== 'non vegetarian']['prep_time'], kde = True, kde_kws = {'linewidth':2}, color = '#b3646f', label = 'Non vegetarian')
ax.axvline(df[df['diet']=='non vegetarian']['prep_time'].mean(), color = 'red')

#Modifying x axis
ax.set_xlim(-5)
ax.set_xlabel('Preperation time')

#Adding legend
ax.legend()

#Adding grid
ax.grid(axis = 'y', alpha = 0.4)

#Adding title and subtitle
fig.text(0.125, 1.03, 'Vegeterian and non vegeterian dish preperation time distribution', fontfamily = 'serif', fontsize = 30, fontweight = 'bold')
fig.text(0.125, 0.95, 'We see that vegetarian dishes take more time to prepare, but non vegetarian dishes are more distributed', fontfamily = 'serif', fontsize = 20)

plt.show()
```

# 4.3 Cook time

```python
fig, ax = plt.subplots(1,1,figsize=(20, 5))

#Adding distribution graph and mean line for vegetarian dishes
sns.distplot(df[df['diet'] == 'vegetarian']['cook_time'], kde = True, kde_kws = {'linewidth': 2}, color = '#aee378', label = 'Vegetarian', bins = 50)
ax.axvline(df[df['diet']=='vegetarian']['cook_time'].mean(), color = 'green')

#Adding distribution graph and mean line for non vegetarian dishes
sns.distplot(df[df['diet']== 'non vegetarian']['cook_time'], kde = True, kde_kws = {'linewidth':2}, color = '#b3646f', label = 'Non vegetarian', bins = 50)
ax.axvline(df[df['diet']=='non vegetarian']['cook_time'].mean(), color = 'red')

#Modifying x axis
ax.set_xlim(-5, 160)
ax.set_xlabel('Cook time')

#Adding legend
ax.legend()

#Adding grid
ax.grid(axis = 'y', alpha = 0.4)

#Adding title and subtitle
fig.text(0.125, 1.03, 'Vegeterian and non vegeterian dish cook time distribution', fontfamily = 'serif', fontsize = 30, fontweight = 'bold')
fig.text(0.125, 0.95, 'We see that vegetarian dishes take more time to cook, but non vegetarian are more distributed', fontfamily = 'serif', fontsize = 20)

plt.show()
```

# 4.4 Flavor profile

```python
#Cheking if everithing is fine with flavor_profile column
df['flavor_profile'].value_counts()
```

There are -1 values in flavor column. I will replace those with "No taste"

```python
#Replacing -1 values with 'no taste'
df['flavor_profile'] = df['flavor_profile'].replace(['-1'], 'No taste')
```

```python
#This code was done similarly as Joshua did

#Counting values
data1 = df['flavor_profile'].value_counts()

#Setting colors for graph
color_map = ['#dfdfdf' for i in range(len(data1))]
color_map[0] = '#d9f640'

#Plottings
fig, ax = plt.subplots(1,1, figsize=(10,5))
ax.bar(data1.index, data1, width = 0.5, edgecolor = 'darkgrey', linewidth=0.6, color = color_map)

#Adding horizontal grid
ax.grid(axis ='y', alpha = 0.4)

#Text
fig.text(0.125, 1.03, 'Flavor count', fontfamily = 'serif', fontsize = 30, fontweight = 'bold')
fig.text(0.125, 0.95, 'We see that spicy is most frequant flavor', fontfamily = 'serif', fontsize = 20)

plt.show()
```

```python
#This code was done similarly as Joshua did

#Counting ratio
flavor_order = df['flavor_profile'].value_counts().index

data_q2q3 = df[['diet', 'flavor_profile']].groupby('flavor_profile')['diet'].value_counts().unstack().loc[flavor_order].fillna(0)
data_q2q3['sum'] = data_q2q3.sum(axis = 1)
data_q2q3_ratio = (data_q2q3.T / data_q2q3['sum']).T[['vegetarian', 'non vegetarian']].sort_values(by = 'vegetarian', ascending = False)[::-1]
data_q2q3_ratio

#Plotting
fig, ax = plt.subplots(1,1, figsize=(10, 5))

ax.barh(data_q2q3_ratio.index, data_q2q3_ratio['vegetarian'], color = '#aee378', label = 'vegetarian')
ax.barh(data_q2q3_ratio.index, data_q2q3_ratio['non vegetarian'], color = '#b3646f', left = data_q2q3_ratio['vegetarian'], label = 'non vegetarian')

#Modifying axis
ax.set_xticks([])
ax.set_xlim(0,1)
ax.set_yticklabels(data_q2q3_ratio.index, fontfamily = 'serif', fontsize = 11)

#Annotating
#For vegetarian part
for i in data_q2q3_ratio.index:
    ax.annotate(f"{round(data_q2q3_ratio['vegetarian'][i]*100, 1)}%",
               xy=(data_q2q3_ratio['vegetarian'][i]/2, i),
               va = 'center', ha = 'center', fontsize = 11, fontfamily = 'serif')

#For non vegetarian part
for i in data_q2q3_ratio.index:
    ax.annotate(f"{round(data_q2q3_ratio['non vegetarian'][i]*100, 1)}%", 
                xy =(data_q2q3_ratio['vegetarian'][i]+data_q2q3_ratio['non vegetarian'][i]/2, i),
               va = 'center', ha = 'center', fontsize = 11, fontfamily = 'serif')

    
#Adding text to graph  
fig.text(0.125, 1.03, 'Vegeterian and non vegeterian flovor ratio', fontfamily = 'serif', fontsize = 30, fontweight = 'bold')
fig.text(0.125, 0.95, 'We see that sweet, bitter and sour dishes are only vegetarian', fontfamily = 'serif', fontsize = 20)

#Adding legend to graph
fig.text(0.76, 0.88, "Vegetarian", fontweight="bold", fontfamily='serif', fontsize=15, color='#aee378')
fig.text(0.9, 0.88, "|", fontweight="bold", fontfamily='serif', fontsize=15, color='black')
fig.text(0.92, 0.88, "Non vegetarian", fontweight="bold", fontfamily='serif', fontsize=15, color='#b3646f')

#Removing spines
for i in ['top', 'bottom', 'left', 'right']:
    ax.spines[i].set_visible(False)

plt.show()
```

```python
#This code was done similarly as Joshua did

#Counting ratio
course_order = df['course'].value_counts().index

data_q3q4 = df[['diet', 'course']].groupby('course')['diet'].value_counts().unstack().fillna(0).loc[course_order]
data_q3q4['sum'] = data_q3q4.sum(axis = 1)
data_q3q4_ratio = (data_q3q4.T/data_q3q4['sum']).T[['vegetarian', 'non vegetarian']].sort_values(by = 'vegetarian', ascending = False)[::-1]

#Plotting
fig, ax = plt.subplots(1,1, figsize=(10, 5))

ax.barh(data_q3q4_ratio.index, data_q3q4_ratio['vegetarian'], color = '#aee378', label = 'vegetarian')
ax.barh(data_q3q4_ratio.index, data_q3q4_ratio['non vegetarian'], left = data_q3q4_ratio['vegetarian'], color = '#b3646f', label = 'non vegetarian')

#Modifying axis
ax.set_xticks([])
ax.set_xlim(0)
ax.set_yticklabels(data_q3q4_ratio.index, fontfamily = 'serif', fontsize = 11)

#Annotating
#For vegetarian
for i in data_q3q4_ratio.index:
    ax.annotate(f"{round(data_q3q4_ratio['vegetarian'][i]*100, 1)}%",
               xy=(data_q3q4_ratio['vegetarian'][i]/2, i),
               va = 'center', ha = 'center', fontsize = 11, fontfamily = 'serif')
    

#For non vegetarian
for i in data_q3q4_ratio.index:
    ax.annotate(f"{round(data_q3q4_ratio['non vegetarian'][i]*100, 1)}%",
               xy = (data_q3q4_ratio['vegetarian'][i]+data_q3q4_ratio['non vegetarian'][i]/2, i),
               va = 'center', ha = 'center', fontfamily = 'serif', fontsize = 11)
    

#Removing spines
for i in ['top', 'right', 'left', 'bottom']:
    ax.spines[i].set_visible(False)
    
#Adding text
fig.text(0.125, 1.03, "Vegeterian and non vegeterian course ratio", fontfamily = 'serif', fontsize = 30,  fontweight = 'bold')
fig.text(0.125, 0.95, 'We see that courses are mostly vegetarian, starter is only non vegetarian', fontfamily = 'serif', fontsize = 20)

#Adding legend
fig.text(0.76, 0.88, "Vegetarian", fontweight="bold", fontfamily='serif', fontsize=15, color='#aee378')
fig.text(0.9, 0.88, "|", fontweight="bold", fontfamily='serif', fontsize=15, color='black')
fig.text(0.92, 0.88, "Non vegetarian", fontweight="bold", fontfamily='serif', fontsize=15, color='#b3646f')

plt.show()
```
