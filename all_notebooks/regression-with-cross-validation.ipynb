```python
import numpy as np 
import pandas as pd
```

```python
gap = pd.read_csv('../input/gapminder/gapminder.csv')
```

```python
gap.head()
```

```python
# CHecking for null values
gap.isna().sum()
```

```python
gap.drop('Region',axis = 1,inplace = True)

```

```python
# Checking columns names
gap.columns
```

```python
gap.describe()
```

```python
x = gap.iloc[:,gap.columns != 'life']
y = gap.loc[:,'life']
```

```python
# importing modeules
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
# dividing the dataset into train and test sets
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state =123)
```

```python
# Creating the Regressor
reg = LinearRegression()
reg.fit(X_train,y_train)
y_pred = reg.predict(X_test)
```

```python
# Finding R-squared
print('R-squared of model :',reg.score(X_test,y_test))
```

```python
# finding root mean squared error
print('Root mean squared error :: ',np.sqrt(mean_squared_error(y_pred,y_test)))
```

Using cross validation

```python
from sklearn.model_selection import cross_val_predict,cross_val_score
```

```python
# cross_val_score gives R-squared values according to cv value
reg_cross_val = LinearRegression()
cross_val_r_squared = cross_val_score(reg_cross_val,x,y,cv = 10)
```

```python
print(cross_val_r_squared)
```

```python
print('mean of cross val_score values::',np.mean(cross_val_r_squared))
```

# Regularisation Regression
## Ridge and Lasso regression

```python
# Implementing Reidge Regression
# importing Ridge
from sklearn.linear_model import Ridge

ridge = Ridge(alpha = 0.1,normalize = True)
ridge.fit(X_train,y_train)
ridge_predict = ridge.predict(X_test)
print('Score using Ridge :: ',ridge.score(X_test,y_test))
```

```python
# Implementing Lasso Regression
# importing Lsso


from sklearn.linear_model import Lasso

lasso = Lasso(alpha = 0.1,normalize = True)
lasso.fit(X_train,y_train)
lasso_predict = lasso.predict(X_test)
print('Score using Lasso :: ',lasso.score(X_test,y_test))
```
