```python
from sklearn.preprocessing import LabelEncoder
var_mod = ['city','toss_decision','venue']
le = LabelEncoder()
for i in var_mod:
    df[i] = le.fit_transform(df[i])
```

```python
df.head(10)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold   
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import metrics

def classification_model(model, data, predictors, outcome):
      model.fit(data[predictors],data[outcome])
      predictions = model.predict(data[predictors])
      print(predictions)
      accuracy = metrics.accuracy_score(predictions,data[outcome])
      print('Accuracy : %s' % '{0:.3%}'.format(accuracy))
```

```python
#Random forest classifier
model = RandomForestClassifier(n_estimators=100)
outcome_var = ['winner']
predictor_var = ['team1', 'team2', 'city','toss_decision','toss_winner','venue']
classification_model(model, df,predictor_var,outcome_var)
```

```python
n = int(input("No. of matches you are going to predict: "))
team1 = []
team2 = []
city = []
toss_decision = []
toss_winner = []
venue = []

for i in range(n):
    user_team1 = input("Enter team 1: ")
    user_team2 = input("Enter team 2: ")
    user_city = input("Enter the city name: ")
    user_tossdecision = input("Enter the toss decision: ")
    user_tosswinner = input("Enter the toss winner: ")
    user_venue = input("Enter the venue: ")
    team1.append(user_team1)
    team2.append(user_team2)
    city.append(user_city)
    toss_decision.append(user_tossdecision)
    toss_winner.append(user_tosswinner)
    venue.append(user_venue)
    
x_new = {'team1':team1,'team2':team2,'city':city,'toss_decision':toss_decision,'toss_winner':toss_winner,'venue':venue}
x_new = pd.DataFrame(x_new)
x_new
```
