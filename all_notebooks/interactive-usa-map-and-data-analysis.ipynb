# As you can see there are a lot of empty records but we'll deal with that later

# I have created a new column that only contains the year

```python
data_kbp['year'] = pd.DatetimeIndex(data_kbp['Date']).year
data_kbp.head()
```

# We create a dataframe containing only year and state(only these columns are needed for our choropleth map)
If you don't know what it is:
* https://en.wikipedia.org/wiki/Choropleth_map
* https://plotly.com/python/choropleth-maps/

```python
states = pd.DataFrame(data_kbp['State'])
states['year'] = pd.DataFrame(data_kbp['year'])
states
```

# Grupby state and year to get the homicide rate for each state by year

```python
state2 = states.groupby(['year', 'State'],as_index=False).size()
state2.rename(columns={'size':'number_of_cases'}, inplace=True)
state2.columns
```

```python
state2
```

```python
import plotly.express as px
```

```python
px.choropleth(state2, 
              locations = 'State',
              color="number_of_cases", 
              color_continuous_scale="Turbo",
              animation_frame='year',
              locationmode='USA-states',
              scope="usa",
              range_color=(0, 250),
              title='Crime by State per Year',
              height=600
             )
```

# Let's go back to cleaning our data
**I deleted the Armed column and then deleted all empty records**

```python
data_kbp=data_kbp.drop(["Armed"],axis=1)
```

```python
data_kbp=data_kbp.dropna()
```

# 10 states with the most cases

```python
top10State = data_kbp.State.value_counts().reset_index(name='Number of deaths per state')
top10State = top10State[:10]
plt.figure(figsize=(10,8))
sns.set_style('darkgrid')
sns.barplot(x='index',y='Number of deaths per state',data=top10State)
```

# 10 cities with the most cases

```python
top10City = data_kbp.City.value_counts().reset_index(name='Number of deaths per City')
top10City = top10City[:10]
plt.figure(figsize=(10,8))
sns.set_style('darkgrid')
sns.barplot(x='index',y='Number of deaths per City',data=top10City)
```
