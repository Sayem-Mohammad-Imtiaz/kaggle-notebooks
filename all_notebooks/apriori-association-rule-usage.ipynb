```python
import pandas as pd
import numpy as np
import seaborn as sns
!pip install mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv("../input/supermarket/GroceryStoreDataSet.csv",names=['Products'],sep=',')
df.head()
```

```python
df.shape
```

```python
df.describe().T
```

```python
data = list(df["Products"].apply(lambda x:x.split(',')))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data,columns=te.columns_)
df
```

```python
freg_items=apriori(df,min_support=0.2,use_colnames=True,verbose=1)
```

```python
#the support value in the first column is the probability that the product (s) specified in column 2 will appear in all sales(the time when the data was received is not specified)
#based on taken data a time in a day,the frequencies can be changed due to sales behaviour will change
freg_items
```

```python
freg_items.sort_values(by="support",ascending=False)
```

```python
df1=association_rules(freg_items,metric="confidence",min_threshold=0.4)
df1
```

*** INFO
* support(X,Y)=Freg(X,Y)/N(Total purchased product)
* confidence(X,Y)=Freg(X,Y)/Freg(X)
* Lift=support(X,Y)/(support(X)*support(Y))



### Some comparisons since support value =0.2,confidence value=0.4

* according to results of biscuit and bread comparison, 57 percent of customers who buy biscuits are observed to buy bread and sales of bread products increase by 87 percent in orders with biscuit same as Tea and Bread association results
* The result of the third case indicates that people who purchase suger will also purchase Coffee. This is followed by a confidence value that shows the possibility of buying Coffee after purchasing suger by %67 apx.
*** As can be seen, the purchase of products affects each other. For this reason, if the specified store is an online store, it would be reasonable to recommend coffee to a customer who buys sugar, or to offer bread to a person who buys milk. If the specified store is not an online store, placing the products bought together in a remote location will allow the customer to look at other products. With the same logic, placing a product that is seen in 10% of shopping like jam next to a product seen in 65% of shopping like bread will allow the customer to take a look at it***
