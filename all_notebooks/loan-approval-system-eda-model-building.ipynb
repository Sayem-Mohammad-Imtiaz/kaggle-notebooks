# Feature Engineering

```python
data.rename(columns={'Sex': 'Male','Education':'Graduation'}, inplace = True)
data.columns
```

```python
data.Male.value_counts().to_frame()
```

```python
data['Male'] = data['Male'].replace({'Male': 1,'Female':0})
```

```python
data.Married.value_counts().to_frame()
```

```python
data['Married'] = data['Married'].replace({'Yes': 1,'No':0})
```

```python
data.head(2)
```

```python
data.Dependents.value_counts().to_frame()
```

```python
data['Dependents']=data['Dependents'].replace({'3+':3})
```

```python
data['Dependents']=data['Dependents'].astype(int)
```

```python
data.Graduation.value_counts().to_frame()
```

```python
data['Graduation']=data['Graduation'].replace({'Graduate':1, 'Not Graduate':0})
```

```python
data.Employee.value_counts().to_frame()
```

```python
data['Employee']=data['Employee'].replace({'Yes':1, 'No':0})
```

```python
data.Area.value_counts().plot(kind='barh')
```

```python
dummies0=pd.get_dummies(data.Area)
```

```python
data= pd.concat([data,dummies0], axis='columns')
data.drop(['Area','Rural'],axis=1, inplace=True)
```

```python
data.dtypes
```

```python
data['Semiurban']=data['Semiurban'].astype(int)
data['Urban']=data['Urban'].astype(int)
```

```python
data['Spouse_Income'].unique()
```

```python
data = data[data.Spouse_Income != '9.857.999.878']
data = data[data.Spouse_Income != '1.612.000.084']
```

```python
data['Spouse_Income']=data['Spouse_Income'].astype(int)
```

```python
data.Approval.unique()
```

```python
data['Approval']=data['Approval'].replace({'Y':1, 'N':0})
```

```python
data.head()
```

```python
x=data.drop({'Client','Approval'}, axis='columns')
y=data['Approval']
```

# Model designing

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import cross_val_score
```

```python
clf0 = LogisticRegression(C=5,solver = 'liblinear')
scores = cross_val_score(clf0, x, y, cv=10)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean()*100, scores.std()))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=9)
```

```python
clf0.fit(x_train,y_train)
clf0.score(x_test,y_test)
```

# Saving the model as a Pickle file.

```python
clf0.fit(x_train,y_train)
Pkl_Filename = "Loan Approval System.pkl"  
import pickle
with open(Pkl_Filename, 'wb') as file:  
    pickle.dump(clf0, file)
print("Task Completed successfully, let Priyatama rest now!")
```
