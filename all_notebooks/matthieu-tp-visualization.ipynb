```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
uscar_filepath = "../input/usa-cers-dataset/USA_cars_datasets.csv"

# Read the file into a variable fifa_data
# uscar_data = pd.read_csv(uscar_filepath, index_col="brand")
uscar_data = pd.read_csv(uscar_filepath)
```

```python
uscar_data
```

```python
#1- Moyenne des prix de vente par Marque

#Premier Filtre : ne montrer que les colonnes Brand et Filtre
uscar_data2=uscar_data.loc[:, ['brand', 'price']]
uscar_data2.head()

#Deuxieme Filtre : Grouper par Brand + moyenne des prix par brand
uscar_data3=uscar_data2.groupby('brand').mean()
uscar_data3

plt.figure(figsize=(30,6))

plt.title("Moyenne des prix par Marque")

sns.barplot(x=uscar_data3.index, y=uscar_data3['price'])

plt.ylabel("Arrival delay (in minutes)")
```

```python
#2-Evolution des prix par ann√©e

#Premier Filtre : ne montrer que les colonnes Brand et Filtre
uscar_data2=uscar_data.loc[:, ['year', 'price']]
uscar_data2.head()

#Deuxieme Filtre : Grouper par Year + somme des prix par an
uscar_data3=uscar_data2.groupby('year').sum()
uscar_data3

sns.regplot(x=uscar_data3.index, y=uscar_data3['price'])
```

```python
#Rapport entre kilometrage et prix
uscar_data2=uscar_data.loc[:, ['brand', 'price','mileage']]
uscar_data2.head()

# uscar_data3=uscar_data2.groupby(['brand']).sum()
uscar_data3=uscar_data2.groupby('brand').agg({'price': 'sum', 'mileage': 'mean'})
uscar_data3

sns.regplot(x=uscar_data3['mileage'], y=uscar_data3['price'])
```
