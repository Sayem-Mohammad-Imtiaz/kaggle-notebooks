```python
tweets_test = pd.read_csv("/kaggle/input/natural-language-processing-with-disaster-tweets/kaggle nlp/test.csv")
tweets_train = pd.read_csv("/kaggle/input/natural-language-processing-with-disaster-tweets/kaggle nlp/train.csv")
tweets_train.info()
```

```python
tweets_train = tweets_train.drop(["location"], axis=1)
tweets_test = tweets_test.drop(["location"], axis=1)
tweets_train.info()
```

```python
vectorizer = CountVectorizer(lowercase=True, stop_words='english', max_features=1000, token_pattern='[a-zA-Z]+')
vectorizer = vectorizer.fit(tweets_train['text'])

bag_of_words_train = vectorizer.transform(tweets_train['text'])
bag_of_words_test = vectorizer.transform(tweets_test['text'])
```

```python
X_train = pd.DataFrame(
    bag_of_words_train.todense(),
    columns=vectorizer.get_feature_names()
)

X_test = pd.DataFrame(
    bag_of_words_test.todense(),
    columns=vectorizer.get_feature_names()
)

X_train.head(20)
```

```python
target_train = tweets_train['target']

from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm

dtc_model = DecisionTreeClassifier(max_depth=3)
dtc_model.fit(X_train, target_train)
predictions = dtc_model.predict(X_train)

#Train
scores = cross_val_score(dtc_model, X_train, target_train, cv=5)
print("Promedio:", dtc_model.score(X_train, target_train))
print("Exactitud de cada particion:", scores)
print("Exactitud Promedio:", scores.mean())
```

```python
gussian_model = GaussianNB()
gussian_model.fit(X_train, target_train)
predictions = gussian_model.predict(X_train)

#Train
scores = cross_val_score(gussian_model, X_train, target_train, cv=5)
print("Promedio:", gussian_model.score(X_train, target_train))
print("Exactitud de cada particion:", scores)
print("Exactitud Promedio:", scores.mean())
```

```python
rfc_model = RandomForestClassifier()
rfc_model.fit(X_train, target_train)
predictions = rfc_model.predict(X_train)

#Train
scores = cross_val_score(rfc_model, X_train, target_train, cv=5)
print("Promedio:", rfc_model.score(X_train, target_train))
print("Exactitud de cada particion:", scores)
print("Exactitud Promedio:", scores.mean())
```

```python
dtr_model = DecisionTreeRegressor(max_depth=2)
dtr_model.fit(X_train, target_train)
predictions = dtr_model.predict(X_train)

#Train
scores = cross_val_score(dtr_model, X_train, target_train, cv=5)
print("Promedio:", dtr_model.score(X_train, target_train))
print("Exactitud de cada particion:", scores)
print("Exactitud Promedio:", scores.mean())
```

```python
svm_model = svm.SVC(kernel = 'rbf')
svm_model.fit(X_train, target_train)
predictions = svm_model.predict(X_train)
predictions_test = svm_model.predict(X_test)

#Train
scores = cross_val_score(svm_model, X_train, target_train, cv=5)
print("Promedio:", svm_model.score(X_train, target_train))
print("Exactitud de cada particion:", scores)
print("Exactitud Promedio:", scores.mean())
```

```python
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(X_train, target_train)
predictions = knn_model.predict(X_train)

#Train
scores = cross_val_score(knn_model, X_train, target_train, cv=5)
print("Promedio:", knn_model.score(X_train, target_train))
print("Exactitud de cada particion:", scores)
print("Exactitud Promedio:", scores.mean())
```
