```python
#--------------LogisticRegression------------------

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt  # import library

dataset = pd.read_csv('../input/social-network-ads/Social_Network_Ads.csv')  # store csv file value in dataset variable

X = dataset.iloc[:, [2,3]].values  # devided feature/dependent
y = dataset.iloc[:, 4].values   # devided prediction/independent

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=0)  # devided train & test(test just few value)

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()   # sc_X here is a variable
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.fit_transform(X_test)  # devided into range of train and test values

from sklearn.linear_model import LogisticRegression   # graph classify & this 2 line are diff of all diagram
classifier = LogisticRegression()

classifier.fit(X_train, y_train)    

y_pred = classifier.predict(X_test)

from sklearn.metrics import confusion_matrix,accuracy_score
cm = confusion_matrix(y_test, y_pred)   # cm variable show how much wrong & right ans give of machine  
ac = accuracy_score(y_test, y_pred)   # given accuracy
print(cm)
print(ac)

# print in a graph
from matplotlib.colors import ListedColormap   
X_set, y_set = X_test, y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 1, stop = X_set[:, 0].max() + 1, step = 0.01),
                     np.arange(start = X_set[:, 1].min() - 1, stop = X_set[:, 1].max() + 1, step = 0.01))
plt.contourf(X1, X2, classifier.predict(np.array([X1.ravel(), X2.ravel()]).T).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Logistic Regression (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
