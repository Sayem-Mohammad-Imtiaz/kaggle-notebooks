```python
from sklearn import preprocessing
```

```python
X = data.drop('class', axis=1) #bize yardım edecek olan kolonlar
y = data['class'] #hedef
y.values
```

```python
#Normalleştirme
x = data[['habitat']].values.astype(float)

min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)

data['habitatN'] = pd.DataFrame(x_scaled)
data.head()
```

```python
#Elimizdeki (X,y) verisini %70 train %30 test şeklinde ayırıyoruz
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
#Modeli eğitiyoruz
LR.fit(X_train,y_train)
```

```python
#Modelin tahmin yeteneğini test ediyoruz
predictions = LR.predict(X_test)
```

```python
#Acc score
score = LR.score(X_test, y_test)
score
```

```python
from sklearn.model_selection import cross_val_predict, cross_val_score
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
#Karmaşıklık matrisi
cm = confusion_matrix(y_test, predictions)
print(cm)
```

```python
#Precision recall vb.
classification_report(y_test,predictions)
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
#Modeli eğitiyoruz
clf.fit(X_train,y_train)
```

```python
#Modelin tahmin yetenegini test ediyoruz
clf_predict = clf.predict(X_test)
```

```python
#ACC score
clf_score = clf.score(X_test, y_test)
clf_score
```

```python
#Karmaşıklık matrisi
confusion_matrix(y_test, clf_predict)
```

```python
#Precision recall vb.
classification_report(y_test,clf_predict)
```

Bir tahminin başarı ölçümünü karmaşıklık matrislerine (confusion matrix) bakarak anlayabiliriz. Tahminin başarılı olabilmesi için recall ve precision değerlerinin olabildiğince yüksek olması gerekir. Bu yüzden iki modeli karşılaştırdığımızda Logistic Regresyonun daha başarılı olduğunu görüyoruz.
