**Logistic Regression using Nerual Netowork**

This is an example to demonstrate logistic reguression with Neural Network using Neural Network 

Dataset: https://www.kaggle.com/mirichoi0218/insurance

Thanks to Daniel Brouke - his youtube channel (@mrdbrouke) is extremly resourceful 

```python
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.model_selection import train_test_split

#Reading the data
insurance = pd.read_csv("../input/insurance/insurance.csv")
insurance.head()
```

```python
#Create X and y values
X = insurance.drop("charges", axis = 1)
y = insurance["charges"]
```

```python
#View X
X.head()
```

```python
#View y
y.head()
```

```python
#Split data into test dataset and training dataset
#Use train_test_split from sklearn

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, train_size = 0.8, random_state = 42)
len(X), len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
#The data should be normalized to improve the results
#Make a column transfromer
ct = make_column_transformer(
    (MinMaxScaler(), ["age", "bmi", "children"]),
    (OneHotEncoder(handle_unknown="ignore"), ["sex", "smoker", "region"])
    )

#Fitting column transformer to training data
ct.fit(X_train)

#Transform training and test data with normalization
X_train_normal = ct.transform(X_train)
X_test_normal = ct.transform(X_test)
```

```python
X_train.loc[0]
```

```python
X_train_normal[0]
```

```python
#Build Neural Network
tf.random.set_seed(42)

#Create model
insurance_model = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

#Compile the model
insurance_model.compile(loss = tf.keras.losses.mae,
                       optimizer = tf.keras.optimizers.Adam(),
                       metrics = ["mae"])

#Fit the model
history = insurance_model.fit(X_train_normal, y_train, epochs = 200, verbose = 1)
```

```python
#Check the results of insurance model on test data
insurance_model.evaluate(X_test_normal, y_test)
```

```python
#Plot history

pd.DataFrame(history.history).plot()
plt.ylabel("loss")
plt.xlabel("epochs")
```
