```python
country_vaccinations.fillna(value=0, inplace=True)
date = country_vaccinations.date.str.split('-', expand =True)
date
```

# Can you trace the daily vaccinations dynamic?
#### China attained the first place for the greatest daily vaccination of its people followed by India and United States. However, countries such as Bhutan	or Falkland Islands	performed better than other countries according to the daily vaccinations per million.
#### According to the tables provided below, Oxford/AstraZeneca	was used more than other vaccines per mellion for daily vaccination program, while CanSino, Sinopharm/Beijing, Sinopharm/Wuhan, Sinovac	were used more than other vaccines for daily vaccination program
