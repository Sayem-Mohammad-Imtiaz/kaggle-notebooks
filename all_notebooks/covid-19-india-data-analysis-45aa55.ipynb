```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def main_menu():
    print("****************************************")
    print("Read Data From File in Different Way")
    print("1.Read complete csv file")
    print("2.Reading complete file without index")
    print("========================================")
    print("Data Visualization")
    print("3.Line Chart")
    print("4.Bar Plot")
    print("5.Pie Plot")
    print("6.Scatter chart")
    print("========================================")
    print("Apply Data Manipulation in the records of CSV file")
    print("7.Sorting the data as per your choice")
    print("8.Read Top and Bottom Records of CSV file")
    print("9.Make the copy of CSV file")
    print("10.Read the Specific columns")
    print("****************************************")
  
main_menu()

def ReadCSV():
    print("/kaggle/input/covid19-india-dataset/covid_india.csv")
    df=pd.read_csv("/kaggle/input/covid19-india-dataset/covid_india.csv")
    print(df)
    
def no_index():
    print("Reading complete file without index value")
    df=pd.read_csv("/kaggle/input/covid19-india-dataset/covid_india.csv",index_col=0)
    print(df)
    
def line_plot():
    df=pd.read_csv("/kaggle/input/covid19-india-dataset/covid_india.csv")
    sn=df['S. No.']
    st=df['Name of State / UT']
    ac=df['Active Cases']
    rc=df['Cured/Discharged/Migrated']
    dth=pd['Deaths']
    cnf=df['Total Confirmed cases']
    plt.xlable("state")
    plt.xtricks(rotation='vertical')
    
    print("Select Specific Line Chart as given below:")
    print("press 1 to print the data for State vs Confirmed Cases")
    print("press 2 to print the data for State vs Recovered Cases")
    print("press 3 to print the data for State vs Death Cases")
    print("press 4 to print the data for State vs Active Cases")
    print("press 5 to merge all the data in one Line Chart")
    
    opt=int(input("Enter your choice:"))
    
    if opt==1:
        plt.ylable("Confimed cases")
        plt.title("State wise Confirmed Cases")
        plt.plot(st,cnf)
        plt.show()
    elif opt==2:
        plt.ylabel("Recovered cases")
        plt.title("State wise Recovered Cases")
        plt.plot(st,rc)
        plt.show()
    elif opt==3:
        plt.ylabel("Death cases")
        plt.title("State wise Death Cases")
        plt.plot(st,dth)
        plt.show()
    elif opt==4:
        plt.ylabel("Active cases")
        plt.title("State wise Active Cases")
        plt.plot(st,ac)
        plt.show()
    elif opt==5:
        plt.ylabel("Number of cases")
        plt.plot(st,cnf,label="State wise Confirmed Cases")
        plt.plot(st,rc,label="State wise Recovered Cases")
        plt.plot(st,dth,label="State wise Death Cases")
        plt.plot(st,ac,label="State wise Active Cases")
        plt.legend()
        plt.show()
    else:
        print("Enter valid input")
    
def bar_plot():
    df=pd.read_csv("/kaggle/input/covid19-india-dataset/covid_india.csv")
    sn=df['S. No.']
    st=df['Name of State / UT']
    ac=df['Active Cases']
    rc=df['Cured/Discharged/Migrated']
    dth=pd['Deaths']
    cnf=df['Total Confirmed cases']
    plt.xlable("state")
    plt.xtricks(rotation='vertical')
    
    
    print("Select Specific Bar Chart as given below:")
    print("press 1 to print the data for State vs Confirmed Cases")
    print("press 2 to print the data for State vs Recovered Cases")
    print("press 3 to print the data for State vs Death Cases")
    print("press 4 to print the data for State vs Active Cases")
    print("press 5 to print all the data in form of stack bar chart")
    print("press 6 to print all the data in form of multi bar chart")
    
    opt=int(input("Enter your choice"))
    
    if opt==1:
        plt.ylable("Confimed cases")
        plt.title("State wise Confirmed Cases")
        plt.bar(st,cnf)
        plt.show()
    elif opt==2:
        plt.ylabel("Recovered cases")
        plt.title("State wise Recovered Cases")
        plt.bar(st,rc)
        plt.show()
    elif opt==3:
        plt.ylabel("Death cases")
        plt.title("State wise Death Cases")
        plt.bar(st,dth)
        plt.show()
    elif opt==4:
        plt.ylabel("Active cases")
        plt.title("State wise Active Cases")
        plt.bar(st,ac)
        plt.show()
    elif opt==5:
        plt.ylabel("Number of cases")
        plt.bar(st,cnf,label="State wise Confirmed Cases")
        plt.bar(st,rc,label="State wise Recovered Cases")
        plt.bar(st,dth,label="State wise Death Cases")
        plt.bar(st,ac,label="State wise Active Cases")
        plt.legend()
        plt.show()
    elif opt==6:
        ind=np.arange(len(st))
        width=0.25
        plt.bar(ind,cnf,width,label="State wise Confirmed Cases")
        plt.bar(ind+0.25,rc,width,label="State wise Recovered Cases")
        plt.bar(ind+0.50,dth,width,label="State wise Death Cases")
        plt.bar(ind+0.75,ac,width,label="State wise Active Cases")
        plt.legend()
        plt.show()
    else:
        print("Enter valid input")
        
def pie_plot():
    df=pd.read_csv("/kaggle/input/covid19-india-dataset/covid_india.csv")
    sn=df['S. No.']
    st=df['Name of State / UT']
    ac=df['Active Cases']
    rc=df['Cured/Discharged/Migrated']
    dth=pd['Deaths']
    cnf=df['Total Confirmed cases']
    
    print("Select Specific Pie Chart as given below:")
    print("press 1 to print the data for State vs Confirmed Cases")
    print("press 2 to print the data for State vs Recovered Cases")
    print("press 3 to print the data for State vs Death Cases")
    print("press 4 to print the data for State vs Active Cases")
    
    opt=int(input("Enter your choice"))
    
    if opt==1:
        plt.title ("State wise Confirmed Cases")
        plt.pie(cnf,labels=st,autopct="%3d%%")
        plt.show()
    elif opt==2:
        plt.title ("State wise Recovered Cases")
        plt.pie(rc,labels=st,autopct="%3d%%")
        plt.show()
    elif opt==3:
        plt.title ("State wise Death Cases")
        plt.pie(dth,labels=st,autopct="%3d%%")
        plt.show()    
    elif opt==4:
        plt.title ("State wise Active Cases")
        plt.pie(ac,labels=st,autopct="%3d%%")
        plt.show()
    else:
        print("Enter valid input")    
        
def scatter_chart():
    df=pd.read_csv("/kaggle/input/covid19-india-dataset/covid_india.csv")
    sn=df['S. No.']
    st=df['Name of State / UT']
    ac=df['Active Cases']
    rc=df['Cured/Discharged/Migrated']
    dth=pd['Deaths']
    cnf=df['Total Confirmed cases']
    
    ax=plt.gca()
    ax.scatter(st,cnf,color='b',lalel="State wise Confirmed Cases")
    ax.scatter(st,rc,color='r',lalel="State wise Recovered Cases")
    ax.scatter(st,dth,color='g',lalel="State wise Death Cases")
    ax.scatter(st,ac,color='y',lalel="State wise Active Cases")
    
    plt.xlabel("state")
    plt.xticks(rotation='vertical')
    plt.title("Complete Scatter chart")
    plt.legend()
    plt.show()
    
def data_sorting():
    df=pd.read_csv("/kaggle/input/covid19-india-dataset/covid_india.csv")
    
    print("Press 1 to arrange the record as per the State Name")
    print("Press 2 to arrange the record as per the Confirmed Cases")
    print("Press 3 to arrange the record as per the Total Deaths")
    print("Press 4 to arrange the record as per the Active Cases")
    
    opt=int(input("Enter your choice"))
    if opt==1:
        df.sort_values(["State"],inplace=True)
        print(df)
    elif opt==2:
        df.sort_values(["Confirmed"],inplace=True)
        print(df)
    elif opt==3:
        df.sort_values(["Recovered"],inplace=True)
        print(df)
    elif opt==4:
        df.sort_values(["Death"],inplace=True)
        print(df)
    elif opt==5:
        df.sort_values(["Active"],inplace=True)
        print(df)
    else:
        print("Enter valid input")
        
def top_bottom_selected_records():
    df=pd.read_csv("/kaggle/input/covid19-india-dataset/covid_india.csv",index_col=0)
    top=int(input("How many records to display from top:"))
    print("First",top,"records")
    print(df.head(top))
    bottom=int(input("How many records to display from bottom:"))
    print("Last",bottom,"records")
    print(df.tail(bottom))
    
def duplicate():
    print("Duplicate the file with new file")
    df=pd.read_csv("/kaggle/input/covid19-india-dataset/covid_india.csv")
    df.to_csv("/kaggle/input/covid19-india-dataset/covid_indiaNew.csv")
    print("Data from the new file")
    print(df)
    
def specific_col():
    print("Reading Specific Column from CSV file")
    df=pd.read_csv("/kaggle/input/covid19-india-dataset/covid_india.csv",usecols=['Name of State / UT','Cured/Discharged/Migrated'],index_col=0)
    print(df)
    
opt=int(input("enter your choice"))
if opt==1:
    ReadCSV()
if opt==2:
    no_index()
elif opt==3:
    line_plot()
elif opt==4:
    bar_plot()
elif opt==5:
    pie_plot()
elif opt==6:
    scatter_chart()
elif opt==7:
    data_sorting()
elif opt==8:
    top_bottom_selected_records()
elif opt==9:
    duplicate()
elif opt==10:
    specific_col()
else:
    print("enter valid input")
    
```
