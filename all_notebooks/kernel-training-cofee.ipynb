İlk önce veriyi bir değişkene atayıp ardından onu "info()" fonksiyonu ile inceleyelim.

First of all, lets assign the data to a variable with using "info()" function and examine the data.

```python
df = pd.read_csv("../input/coffee-quality-database-from-cqi/arabica_data_cleaned.csv")
df.info()
```

Ardından veri tabanındaki sütunların içerdikleri verilerin birbirleri ile olan ilişkisini anlamak için "corr()" fonksiyonunu kullanalarak inceleme yapalım.

After than, to understand the correlation with all columns in the data set, lets use "corr()" function.

```python
df.corr()
```

Bu ilişkiyi daha iyi görebilmek için, yaptığımız işlemi bu sefer ısı haritasında yapalım.

To see the corelation easily, lets use the heat map this time.

```python
f,ax = plt.subplots(figsize = (13,13))
sns.heatmap(df.corr(), annot = True, linewidth = 5, fmt = ".1f", ax=ax )
plt.show()
```

Coffe Arabica tür kahve çekirdeği ile Türk kahvesi yapımı çok yaygındır.
Türk kahvesinde kahvenin gövdesi büyük asiditesinin ise gövdeye göre düşük olması gerek. Isı haritasından bu ilişkiyi görebiliriz. Bunu daha iyi anlamak için tablo ile gösterelim.


Its too common to brew Turkish coffe with Coffe Arabica type coffe.
Turkish coffe should contain high body and low acidity according to body. We can see the correlation in the heat map. So lets use table to see it more clear.

```python
df.Body.plot(kind = "line", color = "Blue", label = "Body", linewidth = 1, alpha = 0.5, grid = True , linestyle = ":")
df.Acidity.plot(color = "red", label = "Acidity", linewidth = 1, alpha = 0.5, grid = True , linestyle = "-.")
plt.legend(loc = "upper right")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line plot")
plt.show()
```

Rakım arttıkça kahvedeki asidite düşer buda Türk kahvesi için önemlidir. Şimdi bunu görelim.

If the altitude increases, acidity will decrease and thats important for Turkish coffe. Now lets see it.

```python
df.altitude_mean_meters.plot(kind = "line", color = "Blue", label = "altitude_mean_meters", linewidth = 1, alpha = 0.5, grid = True , linestyle = ":")
df.Acidity.plot(color = "red", label = "Acidity", linewidth = 1, alpha = 0.5, grid = True , linestyle = "-.")
plt.legend(loc = "upper right")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Line plot")
plt.show()
```

Şimdi ise genel olarak coffe arabica kahve çekirdeğinin tatlı olduğunu grafikler ile gösterelim.

Now lets see coffe arabica coffe flavor rate with graphs.

```python
df.Flavor.plot(kind = "hist", bins = 100, figsize = (15,15))
plt.show
```
