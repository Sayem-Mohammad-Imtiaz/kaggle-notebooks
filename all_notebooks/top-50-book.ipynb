```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../input/amazons-top-50-bestselling-novels-20092020/AmazonBooks - Sheet1.csv')
data.head()
```

```python
sb.lineplot(data=data, x="Year", y="User Rating", hue="Genre",ci=True)
```

```python
df = data
df = df.drop(['Year'], axis=1)
print(df)
```

```python
df = (df.sort_values(by=['User Rating'], ascending=False))
df = df.drop_duplicates()
print(df.head(10))
```

```python
sb.countplot(data= df,x='Genre')
```

```python
df = df.groupby(['Genre', 'User Rating'])['Genre'].count().nlargest(10)
print(df.head(10))
```
