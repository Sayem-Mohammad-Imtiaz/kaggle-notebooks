```python
fatias = [len(df["f_18_20"].value_counts()), len(df["f_70_79"].value_counts())]
labels = ['18 a 20 anos', '70 a 79 anos']
plt.title('Obrigados a votar')
cores  = ['r', 'm']
 
plt.pie(fatias, labels = labels, colors = cores, shadow = False)
 
plt.show()
```
