## Analysing the relationship of petal/sepal lenght/width on the class of flower

```python
#Import python libraries
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
```

**Importing the Heart data**

```python
#Import the heart data
data = pd.read_csv("../input/iris-data/iris_data.csv")
data.head()
```

**Display Information(info) about the data**

```python
data.info()
```

**Display size(shape) about the data**

```python
data.shape
```

**Display basic statistics (describe) of data**

```python
data.describe()
```

**Compare the features 'sepal length (cm)','sepal width (cm)','target' against the target using pairplot**

```python
sns.pairplot(data=data[['sepal length (cm)','sepal width (cm)','target']],hue='target')
```

**Compare the features 'petal length (cm)','petal width (cm)','target' against the target using pairplot**

```python
sns.pairplot(data=data[['petal length (cm)','petal width (cm)','target']],hue='target')
```

<font color='blue'>**OBSERVATION :** Very good correlation exists between target and feature</font>

**Find correlation(corr) of features againts each other**

```python
data.corr()
```

**Find the heatmap of correlation(corr) of features againts each other**

```python
ax = plt.subplots(figsize=(12, 5));
ax = sns.heatmap(data.corr(), annot = True) ;
```

<font color='blue'>**OBSERVATION :** Very good correlation exists between target and feature</font>

**Find the correlation of features against the 'target'**

```python
data.corr()['target'].sort_values(ascending=False)
```

<font color='blue'>**OBSERVATION :** Very good correlation exists between target and feature</font>

**Visualize the 'data' using boxenplot - 'target' vs 'sepal length (cm)'**

```python
sns.boxenplot(x='target', y='sepal length (cm)', data=data)
```

<font color='blue'>**OBSERVATION :**  <br>
    if sepal length is small then it is 'setosa' <br>
    if it is medium then it is 'versicolor'  <br>
    if it is large then it is 'virginica']</font>

**Visualize the 'data' using boxenplot - 'target' vs 'petal length (cm)'**
