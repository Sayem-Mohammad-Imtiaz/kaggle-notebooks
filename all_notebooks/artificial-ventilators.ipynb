```python
nRowsRead = 1000 # specify 'None' if want to read whole file
df = pd.read_csv('../input/cusersmarildownloadsventilatorcsv/ventilator.csv', delimiter=';', encoding = "ISO-8859-1", nrows = nRowsRead)
df.dataframeName = 'ventilator.csv'
nRow, nCol = df.shape
print(f'There are {nRow} rows and {nCol} columns')
```

```python
df.head().style.background_gradient(cmap='summer')
```

```python
cnt_srs = df['quantity'].value_counts().head()
trace = go.Bar(
    y=cnt_srs.index[::-1],
    x=cnt_srs.values[::-1],
    orientation = 'h',
    marker=dict(
        color=cnt_srs.values[::-1],
        colorscale = 'Greens',
        reversescale = True
    ),
)

layout = dict(
    title='Ventilators Distribution',
    )
data = [trace]
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename="quantity")
```

```python
fig = px.pie( values=df.groupby(['quantity']).size().values,names=df.groupby(['quantity']).size().index)
fig.update_layout(
    title = "Quantity of Ventilators by cities",
    font=dict(
        family="Arial, monospace",
        size=10,
        color="#7f7f7f"
    )
    )   
    
py.iplot(fig)
```

```python
fig = px.histogram(df[df.quantity.notna()],x="quantity",marginal="box",nbins=10)
fig.update_layout(
    title = "Quantity of Ventilators",
    xaxis_title="quantity",
    yaxis_title="Number of Ventilators",
    barmode="group",
    bargap=0.1,
    xaxis = dict(
        tickmode = 'linear',
        tick0 = 0,
        dtick = 10),
    font=dict(
        family="Arial, monospace",
        size=15,
        color="#7f7f7f"
    )
    )
py.iplot(fig)
```

```python
sns.countplot(df["quantity"])
plt.xticks(rotation=90)
plt.show()
```

```python
cat = []
num = []
for col in df.columns:
    if df[col].dtype=='O':
        cat.append(col)
    else:
        num.append(col)  
        
        
num 
```

```python
plt.style.use('dark_background')
for col in df[num].drop(['quantity'],axis=1):
    plt.figure(figsize=(8,5))
    plt.plot(df[col].value_counts(),color='Green')
    plt.xlabel(col)
    plt.ylabel('quantity')
    plt.tight_layout()
    plt.show()
```
