```python
import pandas as pd
from pandas.io.json import json_normalize
import warnings
warnings.filterwarnings("ignore")
from sklearn.feature_extraction.text import CountVectorizer
import random
from wordcloud import WordCloud 
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
```

```python
data=pd.read_csv('../input/men-women-shoes-reviews/Shoes_Data.csv')
data
```

```python
data2=data[['reviews','reviews_rating']]
data2
```

```python
data2.info()
```

```python
rating=[]
for item in data2['reviews_rating']:
    rating+=[int(item[0])]
data2['reviews_rating']=rating
data2
```

```python
data2_1=data2[data2['reviews_rating']==1]
data2_2=data2[data2['reviews_rating']==2]
data2_3=data2[data2['reviews_rating']==3]
data2_4=data2[data2['reviews_rating']==4]
data2_5=data2[data2['reviews_rating']==5]
data2s=[data2_1,data2_2,data2_3,data2_4,data2_5,]
```

```python
import re
import nltk
nltk.download("stopwords")
nltk.download('punkt')
from nltk import word_tokenize,sent_tokenize
nltk.download('wordnet')
import nltk as nlp
```

```python
data2s[0]
```

```python
cult_list=[]
for item in data2s:
    df=item['reviews']
    cult_list0=[]
    for cult in tqdm(df):
        cult=re.sub("[^a-zA-z]"," ",str(cult))
        cult=cult.lower()
        cult=nltk.word_tokenize(cult)
        lemma=nlp.WordNetLemmatizer()
        cult=[lemma.lemmatize(word) for word in cult]
        cult=" ".join(cult)
        cult_list0.append(cult)
    cult_list+=[cult_list0]
```

## Normal WordCloud

```python
max_features=200
count_vectorizer=CountVectorizer(max_features=max_features,stop_words="english")
data4=[]
for i in range(5):
    sparce_matrix=count_vectorizer.fit_transform(cult_list[i]).toarray()
    data4+=[pd.DataFrame(count_vectorizer.get_feature_names(),columns=["Words"])]
```

```python
fig,axs = plt.subplots(5,1,figsize=(50,60))
for i in range(5):
    r=i
    random.seed(2021)
    random.shuffle(data4[i].Words)
    wordcloud=WordCloud(background_color="black",width=1024,height=768).generate(" ".join(data4[i].Words))    
    ax=axs[r].axis("off")  
    ax=axs[r].set_title('Rating '+str(i+1))
    ax=axs[r].imshow(wordcloud)
plt.show()
```

### In my opinion, WordCloud of single disjointed words seems to lose the nuance of reviews rating.

# An experimental attempt of **two_word WordCloud**

```python
max_features=200
count_vectorizer=CountVectorizer(max_features=max_features,stop_words="english")
data5=[]
for item in cult_list:
    data5i=[]
    for item2 in item:
        item3=item2.split(' ')
        if len(item3)>1:
            data5i+=[item3[0]+'_'+item3[1]]
    data5+=[data5i]
```

```python
fig,axs = plt.subplots(5,1,figsize=(50,60))
for i in range(5):
    r=i
    random.seed(2022)
    random.shuffle(data5[i])
    wordcloud=WordCloud(background_color="black",width=1024,height=768).generate(" ".join(data5[i]))    
    ax=axs[r].axis("off")  
    ax=axs[r].set_title('Rating '+str(i+1))
    ax=axs[r].imshow(wordcloud)
plt.show()
```

### **Two-word WordCloud** is bad looking due to underbar('_'), but seems to express the nuance of reviews rating more clearly.
