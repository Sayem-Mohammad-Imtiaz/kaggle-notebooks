```python
import pandas as pd
import numpy as np
import string
```

```python
# Reading the data file into a DataFrame and checking the shape
data = pd.read_csv('../input/amazon-mobile-reviews/AmazonMobileDataUncleaned.csv', encoding= "ISO-8859-1")
print(data.shape)
```

```python
data.drop(columns = ['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], inplace=True)
```

```python
# Function to remove punctuations and digits from the text
def removepunctuation(text):
    p = set(string.punctuation) # p take all the punctuation
    text = text.lower() # convert text into lower case
    words = text.split() # split the text into words
    ctext= []
    for i in range(10): #adding numbers from 0 - 9 to p
        p.add(str(i))
    for i in words:
        t = ''.join([x for x in i.encode("ascii", "ignore").decode("ascii") if x not in p])
        ctext.append(t)
    return ' '.join([i for i in ctext]) # join the clean words to text

print(string.punctuation)
uncleaned = [i for i in data.uncleanedreview]
cleaned = []
for i in uncleaned:
    try:
        ctxt = removepunctuation(i)
        if len(ctxt) == 0: raise()
        cleaned.append(ctxt)
    except:
        cleaned.append("NAN")
data['cleantext'] = cleaned
data.to_csv(r'AmazonMobileDataCleaned.csv')
```

```python
# taking the cleaned data and deleting rows that have only special symbols
data = pd.read_csv('./AmazonMobileDataCleaned.csv')
data.drop(data[data['cleantext'] == "NAN"].index, axis = 0, inplace = True)
data = data.reset_index(drop='true')
print(data.shape)
data.to_csv(r'AmazonMobileDataCleanedFinal.csv')
```
