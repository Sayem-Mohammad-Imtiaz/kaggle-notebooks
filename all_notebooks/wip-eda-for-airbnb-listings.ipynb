```python
def loadData(path):
      data = pd.read_csv(path, encoding_errors="ignore")
      return data
```

```python
path = "/kaggle/input/airbnb-listings-reviews/Airbnb Data/Listings.csv"
reviews = loadData(path)
```

### Distribution of unique city count for each host listings

```python
def hostListingAcrossCity(df):
    hostListingCityCount = df.groupby("host_id")["city"].unique().reset_index()
    hostListingCityCount["city_count"] = hostListingCityCount["city"].apply(
        lambda x: len(x)
    )
    print(hostListingCityCount["city_count"].value_counts())
        
hostListingAcrossCity(reviews)
```

> ### For most of hosts, their listings only located in 1 city.
> ### But there are two hosts have listings across 8 cities !!!

## Superhost

### Percentage of superhost by (city, room_type)

```python
plt.figure(figsize=(20, 10))
def superhostByCity(df):
    result = df.groupby(["city", "room_type"]).apply(
        lambda group: (group["host_is_superhost"] == "t").sum() * 100.0 / group.shape[0]
    )
    result = result.to_frame()
    result.reset_index(inplace=True)
    result.rename({0: "Host is superhost (percentage)"}, inplace=True, axis=1)
    sns.barplot(
        x="city", y="Host is superhost (percentage)", hue="room_type", data=result
    )
    plt.show()
    
superhostByCity(reviews)
```
