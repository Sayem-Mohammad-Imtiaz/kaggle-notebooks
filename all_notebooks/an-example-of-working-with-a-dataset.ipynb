```python
import os
import numpy as np
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
dataset_path = '/kaggle/input/synthetic-dataset-for-object-detection/synthetic_dataset/sofa/sofa_v4/'

features = f'{dataset_path}features/images/'
labels = f'{dataset_path}labels/images/'

features_names = os.listdir(f"{features}")
features_names.sort()

labels_names = os.listdir(f"{labels}")
labels_names.sort()

print('features:',len(features_names))
print('labels:',len(labels_names))
```

```python
dataset = []
for f, l in zip(features_names, labels_names):
    f = features + f
    l = labels + l
    dataset.append([f, l])
```

```python
dataset[:3]
```

```python
def show_fl(dup):
    plt.figure(figsize=(10, 10))
    title = ['Image', 'Mask']
    
    feature = cv2.imread(dup[0])
    label = cv2.imread(dup[1], cv2.IMREAD_GRAYSCALE)
    
    dup = (feature, label)
    
    for i in range(len(dup)):
        plt.subplot(1, len(dup), i+1)
        plt.title(title[i])
        plt.imshow(dup[i])
        plt.axis('off')
    plt.show()
    
show_fl(dataset[888])
```
