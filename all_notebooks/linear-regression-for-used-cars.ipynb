```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

cd = pd.read_csv('../input/used-cars-dataset-from-cardekhocom/cardekho_cleaned.csv')

cd.head()
```

```python
# Creating dummy variables for categorical columns

brand_dummies = pd.get_dummies(cd.brand)
model_dummies = pd.get_dummies(cd.model)
seller_dummies = pd.get_dummies(cd.seller_type)
fuel_dummies = pd.get_dummies(cd.fuel_type)
transmission_dummies = pd.get_dummies(cd.transmission_type)
```

```python
merged = pd.concat([cd,brand_dummies,model_dummies,seller_dummies,fuel_dummies,transmission_dummies], axis = 'columns')
merged
```

```python
# Dropping columns for which dummy variables for created
# Dropping one column of dummy variable to avoid dummy variables trap
# Dropping other columns which are not required

merged.drop(['brand','model','seller_type','fuel_type','transmission_type','car_name','new_price','id','Ambassador','Grand','Individual','LPG','Automatic'], axis =1, inplace = True)
```

```python
# imputing the DataFrame for NaN values in columns

from sklearn.impute import SimpleImputer
my_imputer = SimpleImputer()
imputed_merged = my_imputer.fit_transform(merged)
imputed_merged
```

```python
# Converting the above formed array into a DataFrame

merged_df = pd.DataFrame(imputed_merged, columns = merged.columns)
```

```python
X = merged_df.drop(['selling_price'], axis =1)
y = merged_df['selling_price']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

from sklearn.linear_model import LinearRegression

lm = LinearRegression()

lm.fit(X_train,y_train)

predict = lm.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn import metrics

r_2 = r2_score(y_test, predict)
print('The r2 score for the model is',r_2)

print('MAE:', metrics.mean_absolute_error(y_test, predict))
print('MSE:', metrics.mean_squared_error(y_test, predict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predict)))
```
