# Binomial distribution
$P(X=k)=\binom{n}{k}p^k(1-p)^{n-k}$ for $k$ from 1 to $n$.
$E(X)=np$, $V(X)=np(1-p)$.
If $X\sim Bin(n,p)$ and $Y\sim Bin(m,p)$ and independent, then $X+Y\sim Bin(n+m,p)$.
If $n$, mean, and variance are all large, you can approximate with normal.
If $n$ is large and $p$ is small, you can approximate with Poisson with mean $np$.

# Poisson distribution
$P(X=k)=\frac{\lambda^k e^{-\lambda}}{k!}$.
$E(X)=V(X)=\lambda$.
If $X\sim Pois(\lambda)$ and $Y\sim Pois(\mu)$ and independent, then $X+Y\sim Pois(\lambda+\mu)$. 
If $\lambda$ is really large, you can approximate with normal.

# Bernoulli distribution
It's actually binomial with $n=1$. $P(X=1)=p$, $P(X=0)=1-p$. $E(X)=p$. $V(X)=p(1-p)$. 
$P(X=k)=p^k(1-p)^{1-k}$

# Uniform distribution from a to b
It's a continuous distribution.
Density $\frac{1}{b-a}$.
$E(X)=\frac{a+b}{2}$.
$V(X)=\frac{(b-a)^2}{12}$.
