```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn import svm

y =train_data['Survived']

features = ['Pclass','Sex','SibSp','Parch']

X = pd.get_dummies(train_data[features])

X_test = pd.get_dummies(test_data[features])

model = RandomForestClassifier(n_estimators=100,max_depth=5,random_state=1)
model.fit(X,y)

#Try with XGBoost Also
modelXGBoost = svm.SVC()
modelXGBoost.fit(X,y)

#Predictions with both the models
predictionsXGBoost = modelXGBoost.predict(X_test)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId':test_data.PassengerId,'Survived':predictions})
outputXGBoost = pd.DataFrame({'PassengerId':test_data.PassengerId,'Survived':predictionsXGBoost})
outputXGBoost.to_csv('my_submission_svm.csv',index=False)
output.to_csv('my_submissionRF.csv',index=False)
print("Your submission was sucessfully saved...")
```
