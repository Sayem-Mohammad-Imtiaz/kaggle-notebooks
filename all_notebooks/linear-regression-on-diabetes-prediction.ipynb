```python
# The following Notebook shows the application of linear regression on a dataset. Various modules are used here specially linear_model which helps in applying the linear regression.
# Here after reading the dataset u need to alocate the independent columns to a variable and a dependent column to another variable.
# Hence after you have to use train_test_split for training and testing the dataset and finally obtain the predicted values.
# eg: here dependent column is Outcome and rest are independent.
# first lets try out from pregnancies to insulin.

import pandas as pd
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn import model_selection
from sklearn.model_selection import train_test_split

df=pd.read_csv("../input/diabetes-prediction/Diabetespred.csv")
print(df)

x=df.loc[:,'Pregnancies':'Insulin']
y=df['Outcome']

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=4,test_size=0.33)

lm=LinearRegression()
lm.fit(x_train,y_train)
pred=lm.predict(x_test)
print(pred)
print(y_test)
print('Accuracy Score is {}'.format(lm.score(x_train,y_train)))
```
