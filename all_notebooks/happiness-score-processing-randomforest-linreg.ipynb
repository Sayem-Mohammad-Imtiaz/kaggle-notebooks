## Linear regression

```python
LinReg = RandFor.reset_index().drop(['R', 'R_codes', 'index'], axis = 1)
LinReg.columns
```

```python
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score

X = LinReg[['Economy', 'Family', 'Freedom', 'Generosity', 'Health', 'Trust','Year']]
y = LinReg['HS']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

regr = linear_model.LinearRegression()
regr.fit(X_train, y_train)
```

```python
y_pred = regr.predict(X_test)
```

```python
print('Coefficients: \n', regr.coef_)
# The mean squared error
print('Mean squared error: %.2f'
      % mean_squared_error(y_test, y_pred))
# The coefficient of determination: 1 is perfect prediction
print('Coefficient of determination: %.2f'
      % r2_score(y_test, y_pred))
```

```python
RegCoefs = pd.DataFrame(X.columns, columns = ['KPI'])
RegCoefs['coefficient'] = pd.DataFrame(regr.coef_)
RegCoefs.sort_values(by = 'coefficient', ascending = False)
```
