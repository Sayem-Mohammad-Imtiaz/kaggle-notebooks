Unemployement rate is calculated based on region,where due to covid-19 the rate of unemployement increases more.let us analysis using python

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data=pd.read_csv('../input/unemployement-data-till-11-2020/unemp_upto_11_2020.csv')
```

some of the column names are not correct.so its better to rename it

```python
data.columns=["states","date","frequency","Estimated un emp rate","Estimated emp"," Estimated Labour Participation Rate","region","longitude","latitude"]
```

```python
data.columns
```

let us see the correlation blw features

```python
sns.heatmap(data.corr(),annot=True)
```

let us see unemployement across region wise

```python
unemploment = data[["states", "region", "Estimated un emp rate"]]
figure = px.sunburst(unemploment, path=["region", "states"], 
                     values="Estimated un emp rate", 
                     width=700, height=700, color_continuous_scale="RdY1Gn", 
                     title="Unemployment Rate in India")
figure.show()
```

Conclusion: due to Covid-19 above shown % of unemployment leads to increase in every state
