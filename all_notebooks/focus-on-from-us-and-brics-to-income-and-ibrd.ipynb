# 3) Trial to find other factors

First, I tried to find countries or areas where there are large increase in transporation by numbers.

```python
df1.head()
```

```python
for df in [df1, df2, df3, df4]:
    df['Increase'] = df['2019'] - df['2005']
    df['Increase%'] = df['Increase'] / df['2005'] * 100
    df.sort_values('Increase', ascending=False, inplace=True)
```

```python
df1_1=df1.loc[:,['Country Name','Increase','Increase%']]
df2_1=df2.loc[:,['Country Name','Increase','Increase%']]
df3_1=df3.loc[:,['Country Name','Increase','Increase%']]
df4_1=df4.loc[:,['Country Name','Increase','Increase%']]
df1_1 = df1_1.set_index('Country Name')
df2_1 = df2_1.set_index('Country Name')
df3_1 = df3_1.set_index('Country Name')
df4_1 = df4_1.set_index('Country Name')
```

```python
df1_1['Increase'].head(20).plot.bar()
plt.title('Air transport, registered carrier departures worldwide')
```

```python
df2_1['Increase'].head(20).plot.bar()
plt.title('Air transport, freight (million ton-km)')
```

```python
df3_1['Increase'].head(20).plot.bar()
plt.title('Air transport, passengers carried')
```

According to the above graphs, I understand that 'IDA & IBRD countries' or 'Middle income countries' have been contributing to the increase of transportation.
'IDA & IBRD' are categorized by 'lending' from WORLD BANK. And 'Middle income countries' are categorized by income. 
According to the website of WORLD BANK, both countries which are categorized in 'IDA & IBRD countries' and 'Middle income countries' are similar very much.
https://datahelpdesk.worldbank.org/knowledgebase/articles/906519-world-bank-country-and-lending-groups

It means that the money from WORLD BANK have been boosting the economy of countries from low income to middle income. And as a result, it has been contributing to the increase of moving of products and people.
