```python
def ss(d):
    k = []
    k = d.split('-')
    return int(k[0])

df["Day"] = df['Date'].apply(ss)
df['Year'] = pd.DatetimeIndex(df['Date']).year
df['Month'] = pd.DatetimeIndex(df['Date']).month
```

```python
df1['STD'] = df1['Store'].apply(lambda x : df[df['Store'] == x]['Weekly_Sales'].std())
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df1['Store'],y=df1['STD'])
plt.tight_layout()
```

Store #14 has max std

```python
def goodQ312(d):
    f = df[(df['Store'] == d) & (df['Year'] == 2012) & (df['Month']>6) & (df['Month']<10)]['Weekly_Sales'].sum()
    p = df[(df['Store'] == d) & (df['Year'] == 2012) & (df['Month']>3) & (df['Month']<7)]['Weekly_Sales'].sum()
    return ((f-p)/p)
df1['GrowthR'] = df1['Store'].apply(goodQ312)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=df1['Store'],y=df1['GrowthR'])
plt.tight_layout()
```

Growth rate w.r.t past quarter

Store #16 best growth rate
