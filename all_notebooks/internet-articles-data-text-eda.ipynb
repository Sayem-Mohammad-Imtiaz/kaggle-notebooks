```python
def clean_text(text, lemmatizer = WordNetLemmatizer(), 
                  stop_words = set(stopwords.words('english'))):
      
    words = word_tokenize(text)
    
    filtered_words = []
    
    for word in words:
        
        if word not in stop_words and word.isalpha():
            filtered_words.append(lemmatizer.lemmatize(word))
    
    return filtered_words
```

# Title Text analysis

```python
title_word_list = df['title'].apply(lambda x: clean_text(x))
```

```python
title_word_list[:10]
```

```python
word_length_title = [len(w) for w in title_word_list]
```

```python
plt.hist(word_length_title, bins=25)
plt.title('WordLength for title')
plt.ylabel('count')
plt.xlabel('Length of text')
plt.grid()
plt.show()
```

```python
words_title = clean_text(''.join(str(df['title'].tolist())))
```

```python
words_title[:15]
```

```python
(pd.Series(nltk.ngrams(words_title, 2)).value_counts())[:20]
```

```python
bigrams_series_title = (pd.Series(nltk.ngrams(words_title, 2)).value_counts())[:20]

bigrams_series_title.sort_values().plot.barh(color='blue', width=.9, figsize=(12, 8))
plt.title('20 Most Frequently Occuring Bigrams')
plt.ylabel('Bigram')
plt.xlabel('# of Occurances')
```

```python
word_cloud = WordCloud(background_color="white", 
               max_words=2000,
               relative_scaling='auto',
               colormap='winter')

word_cloud.generate(','.join(words_title))
plt.figure( figsize=(12,10) )
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
