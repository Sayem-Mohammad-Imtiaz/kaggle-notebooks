Близкое к 0 значение может говорить о том, что оба эти показателя распределены нормально. Проверим на нормальность данные, описывающие уровень рН:

Построим вариационный ряд. Возьмем 10 интервалов:

```python
t = Target
varr = pd.DataFrame()
varr['2.74-2.86'] = [len(t[(2.74 <= t) & (t <= 2.86)])]
varr['2.87-2.93'] = [len(t[(2.87 <= t) & (t <= 2.93)])]
varr['2.94-3.08'] = [len(t[(2.94 <= t) & (t <= 3.08)])]
varr['3.09-3.19'] = [len(t[(3.09 <= t) & (t <= 3.19)])]
varr['3.2-3.3'] = [len(t[(3.2 <= t) & (t <= 3.3)])]
varr['3.31-3.42'] = [len(t[(3.31 <= t) & (t <= 3.42)])]
varr['3.43-3.56'] = [len(t[(3.43 <= t) & (t <= 3.56)])]
varr['3.57-3.7'] = [len(t[(3.57 <= t) & (t <= 3.7)])]
varr['3.71-3.8'] = [len(t[(3.71 <= t) & (t <= 3.8)])]
varr['3.81-3.89'] = [len(t[(3.81 <= t) & (t <= 3.89)])]
varr
```

Далее построим гистограмму:

```python
plt.figure(figsize = (25, 5))
plt.bar(height = varr.iloc[0], x = varr.columns);
```

Изобразим эмпирическую плотность:

```python
plt.figure(figsize = (15, 5))
plt.plot(varr.iloc[0]);
```

Далее посчитаем асимметрию и эксцесс: первым делом найдем выборочное среднее - просумируем данные и поделим на длину выборки:

```python
n = len(Target)
Srednee = sum(Target)/n
Srednee
```

Таким образом, среднее выборки - 3.311113195747343
Далее посчитаем среднеквадратическое отклонение $\sigma$ - для этого вычтем из каждого значения среднее, возведем в квадрат, просуммируем и поделим на объем выборки

```python
Sigma = (sum((Target - Srednee)**2)/n)**(1/2)
Sigma
```

Таким образом, $\sigma$ = 0.15433818141060152. Теперь можно посчитать ассиметрию:

```python
Assimetria = sum((Target - Srednee)**3)/(n*Sigma**3)
Assimetria
```

Значит, асимметрия равна 0.1935017589100295. Найдем эксцесс:

```python
Ekscess = sum((Target - Srednee)**4)/(len(Target)*Sigma**4)-3
Ekscess
```

Эксцесс получился равным 0.8006714441211811. Далее найдем их теоретические дисперсии:

```python
Dispersia_assim = 6*(n-1)/((n+1)*(n+3))
print(Dispersia_assim)
Dispersia_eks = 24*n*(n-2)*(n-3)/((n+3)*(n+5)*(n+1)**2)
print(Dispersia_eks)
```

По результатам расчетов - дисперсия асимметрии равна 0.00374063670411985 и дисперсия эксцесса равна 0.014869304924136288. Сравним полученные резульаты:

```python
print('Асимметрия меньше либо равна трех корней из диспресии асимметрии - это: ', abs(Assimetria)<=3*Dispersia_assim**(1/2))
print('Эксцесс меньше либо равен пяти корней из диспресии ексцесса - это: ', abs(Ekscess)<=5*Dispersia_eks**(1/2))
```

Поскольку оба ответа False, то это означает, что полученные результаты значительно отклоняются от нуля.

Далее построим интервальные оценки. Для начала -  для математического ожидания, для этого посчитаем точечную несмещенную оценку дисперсии:

```python
S = ((Sigma**2)*n/(n+1))**(1/2)
S
```

Среднее выборки у нас подсчитано ранее, поэтому осталось выбрать уровень значимости и найти соответствующее значение распределения Стьюдента по таблице. Пусть уровень значимости будет равен 0.9, тогда при таком параметре значение распределения Стьюдента будет равным 0.81524, далее найдем доверительный интервал:

```python
print(Srednee - (S/(n**(1/2)))*0.81524, ' ; ', (Srednee + (S/(n**(1/2)))*0.81524))
```

Далее построим доверительный интервал для диспресии с уровнем значимости 0.9. Для этого по таблице найдем значение 

q = q (γ, n) = q(0.9, 155) = 0.111, тогда доверительный интервал имеет вид:

```python
print(S*(1-0.111),' ; ', S*(1+0.111))
```

Вывод: при уровне значимости 0.9 среднее находится в пределах (3.3079676292720137  ;  3.3142587622226722), а среднеквадратическое отклонение в пределах (0.13716375949636406  ;  0.17141612688465746). 

Сформулируем гипотезу о нормальности распределения: при уровне значимости 0.05 проверим гипотезу о том, что данные распределены нормально. При таком уровне значимости  проверим гипотезу  о нормальном распределении $H_0$ генеральной совокупности против конкурирующей гипотезы $H_1$  о том, что она так не распределена. Пользоваться будем критерием Пирсона.

```python
t = varr.T.rename(columns = {0:'N_i'})
t['X_i'] = [round((float(list(varr.columns)[i].split('-')[0]) + float(list(varr.columns)[i].split('-')[1]))/2, 3) for i in range(len(varr.columns))]
t['X_i*N_i'] = t['N_i']*t['X_i']
t['(X_i^2)*N_i'] = t['N_i']*t['X_i']*t['X_i']
t
```

```python
import math
t = varr.T.rename(columns = {0:'N_i'})
t['X_i'] = [round((float(list(varr.columns)[i].split('-')[0]) + float(list(varr.columns)[i].split('-')[1]))/2, 3) for i in range(len(varr.columns))]
t['Z_i'] = (t['X_i'] - Srednee)/S
t['f(Z_i)'] = (math.e**(-(t['Z_i']**2)/2))/((2*math.pi)**(1/2))
Ni = []
for i in range(len(t)):
    Ni.append(t['f(Z_i)'][i]*(sum(t['N_i']))*1/S*(float(t.index[0].split('-')[1]) - float(t.index[0].split('-')[0])))
t["N_i'"] = Ni
t
```

Далее воспользуемся критерием согласия Пирсона. Поскольку всего имеется 10 интервалов, то количество степеней свободы равно 7. Значит, табличное значение $\chi^2 = 14.1$. Вычислим наблюдаемые значения предоставленных данных:

```python
t['Хи^2'] = ((t['N_i'] - t["N_i'"])**2)/t["N_i'"]
t
```

Просуммируем последний столбец: 

```python
sum(t['Хи^2'])
```

Значение  в 11.7 единиц не превышает ожидаемое 14.1, поэтому отвергнуть гипотезу $H_0$ у нас нет оснований.
