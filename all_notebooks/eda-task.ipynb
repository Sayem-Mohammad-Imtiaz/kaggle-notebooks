```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/cvdcvd-vd/Social_Network_Ads.csv')
df.head()
```

```python
df = df.drop('User ID', axis=1)
```

```python
plt.figure(figsize=(9,4))
sns.countplot(x='Purchased', data = df)
```

```python
plt.figure(figsize=(9,4))
sns.countplot(x='Purchased', data = df[df['EstimatedSalary']>df['EstimatedSalary'].mean()])
```

```python
plt.figure(figsize=(9,4))
sns.countplot(x='Purchased', data = df[df['EstimatedSalary']<=df['EstimatedSalary'].mean()])
```

clearly it can be seen that if your salary is more than the mean of the salary data, you are more likely to but the product on seeing the ad. 

```python
plt.figure(figsize=(9,9))
sns.scatterplot(x='Age', y ='EstimatedSalary', hue = 'Purchased', data=df)
plt.plot([i for i in range(62)], [df['EstimatedSalary'].mean() for j in range(62)], 'r--')
plt.plot([df['Age'].mean()for i in range(len(df['EstimatedSalary'].values))],df['EstimatedSalary'].values, 'y--')
```
