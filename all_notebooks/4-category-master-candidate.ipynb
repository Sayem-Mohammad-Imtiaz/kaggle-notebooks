```python
import pandas as pd


UserAchievements = pd.read_csv('/kaggle/input/meta-kaggle/UserAchievements.csv')
Users = pd.read_csv('/kaggle/input/meta-kaggle/Users.csv')
```

```python
# 0: Novice, 1: Contributor, 2: Expert, 3: Master, 4: Grandmaster
UserAchievements['Tier'].value_counts()
```

```python
UserAchievements.query('Tier >= 3')['AchievementType'].unique()
```

```python
agg_df = UserAchievements.query('Tier >= 3').groupby('UserId').count()['AchievementType'].reset_index()
candidate_id = agg_df.query('AchievementType == 3')['UserId'].values
```

```python
df = Users[Users['Id'].isin(candidate_id)].reset_index(drop=True)
df['Url'] = ['https://www.kaggle.com/' + d for d in df['UserName']]
df
```
