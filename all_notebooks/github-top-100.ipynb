```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('/kaggle/input/most-starred-github-repositories/Most starred Github Repositories.csv')
```

```python
df.info()
```

```python
df.convert_dtypes().head()
```

```python
df['Language'].unique()
```

```python
sns.set_style('darkgrid')

fig, ax = plt.subplots(figsize = (15,10))

df['Language'].value_counts().plot(kind = 'bar', color = 'purple', alpha = 0.7)
plt.xticks(rotation = 45)
plt.ylabel('Number of repositories')
plt.xlabel('Language')
plt.title('Most used Languages in the Github Top 100', fontsize = 18)

plt.show()
```

```python
X = df['Language']
y = df['Stars']


fig, ax = plt.subplots(figsize = (15,10))

plt.scatter(X, y, s = 90, alpha = 0.7)
plt.xticks(rotation = 45)
plt.ylabel('Number of Stars')
plt.xlabel('Language')
plt.title('Number of Stars by Repositorie', fontsize = 18)

plt.show()
```

```python
y = df['Forks']


fig, ax = plt.subplots(figsize = (15,10))

plt.scatter(X, y, s = 90, alpha = 0.7)
plt.xticks(rotation = 45)
plt.ylabel('Number of Forks')
plt.xlabel('Language')
plt.title('Number of Forks by Repositorie', fontsize = 18)

plt.show()
```

## Results

- As shown in the first figure, the most used language in the **Github Top 100** is JavaScript, followed by Python and Java. The repositories with no language are guides and books kind and they take the second place in the most common repositries in the dataset.
- There are 20 different languages, and eleven of them got in the top 100 with only one repository.
- The most starred repositorie is written in JS and the most forked one is C++ as shown in the scatter plots
