```python
data=pd.read_csv('/kaggle/input/netflix-original-films-imdb-scores/NetflixOriginals.csv')
data.info()
```

```python
#changing the column name to an easier accessible one
datanew = data.rename(columns={'IMDB Score': 'IMDB'})
datanew.info()
```

```python
datanew.corr()
```

```python
#correlation map
f,ax=plt.subplots(figsize=(10,10))
sns.heatmap(datanew.corr(),annot=True,linewidths=.5,fmt='.1f',ax=ax)
plt.show()
```

```python
datanew.columns
```

```python
#Line Plot
#This line plot shows us the distrubiton of Runtime&IMDB values.
datanew.Runtime.plot(kind = 'line', color = 'brown',label = 'Runtime',linewidth=2,alpha = 1,grid = True,linestyle = 'dotted')
datanew.IMDB.plot(color = 'blue',label = 'IMDB',linewidth=2, alpha = 1,grid = True,linestyle = '-.')
plt.legend(loc='upper right') # legend = puts label into plot    
plt.show()
```

```python
#Histogram Plot
#This histogram plot shows us the frequency of IMDB values
datanew.IMDB.plot(kind='hist',bins=50,figsize=(12,12),color='pink')
plt.show()
```

```python
#Scatter Plot
#This scatter plot shows us the distrubiton of Runtime&IMDB values.
datanew.plot(kind='scatter',x='Runtime',y='IMDB',color='yellow')
plt.show()
```

```python
#Filtering pandas data frame
#Filtering movies with IMDB scores greater than 8 gives us 16 movies.
filter1=datanew.IMDB>8
datanew[filter1]
```

```python
#Method: filtering pandas with logical_and
#Filtering movies whose IMDB is greater than 8, and whose language is Spanish gives us only one piece
datanew[np.logical_and(datanew['IMDB']>8,datanew['Language']=='Spanish')]
```

```python
#Filtering movies whose IMDB is greater than 7 and whose genre is animation gives us 3 pieces:
datanew[np.logical_and(datanew['IMDB']>7,datanew['Genre']=='Animation')]
```

```python
avg=sum(datanew.IMDB)/len(datanew.IMDB)
datanew['quality_level']=['successful' if i>avg else 'unsuccessful' for i in datanew.IMDB]
datanew
```
