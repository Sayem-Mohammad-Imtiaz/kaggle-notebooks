```python
import matplotlib.pyplot as plt

import seaborn as sns
def plot_missing_data(dataset, title):
    fig, ax = plt.subplots(figsize=(5,5))
    plt.title(title)  
    sns.heatmap(dataset.isnull(), cbar=False)
```

```python
plot_missing_data(dataset, "Kerala Assembly Election History")
```

```python
def display_counts(dataset, feature, title, fsize = (5,5)):
    fig, ax = plt.subplots(figsize=fsize)
    sns.countplot(dataset[feature], ax=ax).set_title(title)
```

```python
display_counts(dataset, "Type", "Distribution by Type")
```
