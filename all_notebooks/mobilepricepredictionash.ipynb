```python
#Standardize the data
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
```

```python
y=data_train['price_range']
y
```

```python
X=data_train.drop('price_range',axis=1)
```

```python
data_test.drop(['id'],axis=1,inplace=True)
```

```python
X_std=std.fit_transform(X)
data_test_std=std.transform(data_test)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_std,y)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
dataTest_pred=lr.predict(data_test_std)
dataTest_pred
```

```python
#We cannot find accuracy score,as we do not have ground truth
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_std,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
y_test
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
y_predDT=dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
dt_acc=accuracy_score(y_predDT,y_test)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
y_predKNN=knn.predict(X_test)
```

```python
knn_acc=accuracy_score(y_predKNN,y_test)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_predlr=lr.predict(X_test)
```

```python
lr_acc=accuracy_score(y_predlr,y_test)
```

```python
import matplotlib.pyplot as plt
plt.bar(x=['dt','knn','lr'],height=[dt_acc,knn_acc,lr_acc])

plt.xlabel("Algorithms")
plt.ylabel("Accuracy Score")
plt.show()
```
