```python
import numpy as np

corr = df.corr()
mask = np.triu(np.ones_like(corr,dtype = bool))

plt.figure(figsize = (12,12),dpi = 80)
plt.title('Correlation Analysis')
sns.heatmap(df.corr(),mask=mask,annot=False,cmap='viridis')
plt.xticks(rotation=90)
plt.yticks(rotation = 0)
plt.show()
```
