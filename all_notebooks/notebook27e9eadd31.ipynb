```python
data8=data[["Location","Rainfall"]]
data10=data8.groupby(["Location","Rainfall"]).size()
canberra_mean=data10["Canberra"].mean()
sydney_mean=data10["Sydney"].mean()
hobart_mean=data10["Hobart"].mean()
data_new={"Location":["Canberra","Sydney","Hobart"],
         "Rainfall_mean": [canberra_mean,sydney_mean,hobart_mean]
         }
df=pd.DataFrame(data_new,columns=["Location","Rainfall_mean"])
df.plot(x="Location",y="Rainfall_mean",kind="bar")
```
