## Introduction

> Context:

London's fire and rescue service is the busiest in England and one of the largest firefighting and rescue organisations in the world. In the aftermath of the Grenfell Tower fire, it is critical that firefighting resources are accurately and appropriately deployed.

> Content:

This data covers Jan 01-April 30 2017, consisting of 32 columns containing information on time, type, and address of call, as well the home station, stay duration, and arrival time of attending pumps.

> Inspiration:

Which boroughs have the shortest average call response? Longest?
Which boroughs have the greatest volume of calls?

## Importing and data prep

```python
# Import libraries
import pandas as pd
pd.options.display.max_columns = 999
import numpy as np
import seaborn as sns
from plotnine import *
import re
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
# Import the data set
london = pd.read_csv('../input/london_fire_brigade_service_calls.csv',
    infer_datetime_format=True,
    parse_dates=['timestamp_of_call', 'date_of_call'], )

london['date_of_call'] = london['date_of_call'].dt.date
```

#### Feature Creation

```python
#Feature Creation for response time in minutes
london['First_response_mins']= london['first_pump_arriving_attendance_time']/60
london['Second_response_mins']= london['second_pump_arriving_attendance_time']/60
#Date with year-month format
london['month'] = pd.to_datetime(london['date_of_call']).dt.month
```

## Data Exploration and Visualization

### Trend of Calls between Jan-April 2017
* There is a clear outlier in the day in **Feb 2017**.
* Feb 23 had **525 calls**, which the highest. The day prior Feb 22 had the **lowest in the data at 195**. Interesting that the highest and the lowest are right next to each other. 

```python
(ggplot(data=london) + aes(x= 'date_of_call', group=1) + stat_count(geom='line') + theme_538() +theme(figure_size=(15,7))
+ scale_x_date(limit=["2017-01-01","2017-04-30"]))
```

```python
london.groupby(['date_of_call']).size().sort_values(ascending=False)
```

### Call trend by the hour of day
* The number of calls increases from 0500 and reaches its peak at 1800.
* Early morning hours have understandably lower volume

```python
(ggplot(data=london) + aes(x='hour_of_call',group=1) + geom_bar() + ggtitle(title='Calls by hour of day')
 + theme_538())
```

### How does the average first response time vary by hour of the call?
* Regardless of the hour of call, the average first response time is around 5 mins for all hours of the day. This is within the department's target of 6 mins^. 
* After 2000 hrs, the average first response time declines

^ Source: http://www.london-fire.gov.uk/Documents/London-Fire-Brigade-Our-Performance-2016-17.pdf

```python
plt.figure(figsize=[15,7],)
plt.title('Average First Response Time (mins) by Hour of Call')
sns.swarmplot(data=london,x='hour_of_call', y='First_response_mins')
```

```python
plt.figure(figsize=[15,7],)
plt.title('Average First Response Time (mins) by Hour of Call')
sns.boxplot(data=london,x='hour_of_call', y='First_response_mins')
plt.plot([-1,24],[6, 6], linewidth=3,linestyle='dashed', )
```

### Average Second Response Rate
* The Second Response Rate is also well within the target of 8 mins set by the Fire Department.
* The average response time rise between 1200 to 1500 and then begins to decline.

```python
plt.figure(figsize=[15,7],)
plt.title('Average Second Response Time (mins) by Hour of Call')
sns.boxplot(data=london,x='hour_of_call', y='Second_response_mins')
plt.plot([-1,24],[8, 8], linewidth=3,linestyle='dashed', )
```

### Different Types of Incidents

#### Let's take a look at the number of different types of incidents
* The most number of calls were from AFA or Automatic Fire Alarm
* Special Services are non-emergency situations.


```python
london.groupby(['stop_code_description']).size().sort_values().plot(kind='barh', figsize=[8,4], title='Calls by Incident Type')
```

#### Let's understand the first and second response time for these
* The first response is **fastest for AFA**. We ignore Chimney because of the limited data points.

```python
plt.figure(figsize=[15,7],)
plt.title('Average First Response Time (mins) by Incident Type')

ranks = london.groupby("stop_code_description")["First_response_mins"].mean().fillna(0).sort_values()[::-1].index
sns.boxplot(data=london, y='stop_code_description', x='First_response_mins', orient='h', order=ranks)
```

#### Second Response Time

```python
plt.figure(figsize=[15,7],)
plt.title('Average Second Response Time (mins) by Incident Type')

ranks = london.groupby("stop_code_description")["Second_response_mins"].mean().fillna(0).sort_values()[::-1].index
sns.boxplot(data=london, y='stop_code_description', x='Second_response_mins', orient='h', order=ranks)
```

### Special Service Type

#### Number of Calls

```python
london[london['stop_code_description'] == 'Special Service'].groupby(
    ['special_service_type']).size().sort_values()
```

#### Special Service Type & Average First Response
* First response time is **longest** for **Animal assistance incidents**.

```python
plt.figure(
    figsize=[15, 8], )
plt.title('Average First Response Time (mins) by Special Service Type')

ranks = london[london['stop_code_description'] == 'Special Service'].groupby([
    "special_service_type"
])["First_response_mins"].mean().fillna(0).sort_values()[::-1].index

sns.boxplot(
    data=london[london['stop_code_description'] == 'Special Service'],
    y='special_service_type',
    x='First_response_mins',
    orient='h',
    order=ranks, )
```

### Borough

#### Borough and Fire Brigade Calls

```python
london.groupby(['borough_name']).size().sort_values().plot(kind='barh', figsize= (14,7))
```

#### Average first response time by Borough
* Kingston and Chelsea has the **lowest** average first response time.
* Hillingdon has the highest

```python
plt.figure(
    figsize=[12, 10], )
plt.title('Average First Response Time (mins) by Borough')

ranks = london.groupby(["borough_name" ])["First_response_mins"].mean().sort_values()[::-1].index
sns.boxplot(
    data=london,
    y='borough_name',
    x='First_response_mins',
    orient='h',
    order=ranks,)
```

#### Borough and Call Trend
* In most boroughs, February has a lower number of calls.
