The last four features are not binary features and not highly correlated and hence we will not use them for prediction.

### Feature and Target selection

```python
y=df_noinf['PCOS (Y/N)']
X=df_noinf[['Insulin levels (æIU/ml)', 'Follicle No (R)','Follicle No (L)']]
```

### Test and Train data 

```python
X_train,X_test,y_train, y_test=train_test_split(X,y, test_size=0.2)
```

### Train Model

```python
model=LogisticRegression()
model.fit(X_train,y_train)
print(f"Score in Train Data : {model.score(X_train,y_train)}")
```

### Prediction

```python
y_pred=model.predict(X_test)
```

### Model Evaluation

```python
print(f"Score in Test Data : {model.score(X_test,y_test)}")

cm=confusion_matrix(y_test, y_pred)
p_right=cm[0][0]+cm[1][1]
p_wrong=cm[0][1]+cm[1][0]

print(f"Right classification : {p_right}")
print(f"Wrong classification : {p_wrong}")
cm
```
