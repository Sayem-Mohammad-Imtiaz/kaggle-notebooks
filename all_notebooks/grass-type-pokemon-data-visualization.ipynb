We have a pokemon dataset. We will see the general lay of the data, and find the pokemon with exceptional stats and visualize their statistics in graphs

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pll
```

```python
df=pd.read_csv('/kaggle/input/pokemon/Pokemon.csv')
df.head(0)
```

Seperating grass type pokemon, filtering out all non-mega from them and deleting the pokedex number (#).

```python
gr_df=df.loc[(df['Type 1']=='Grass') & (~df['Name'].str.contains('Mega'))];
del gr_df['#'];
gr_df.plot.line()
gr_df.reset_index();
```

We then plotted the whole data frame as a line chart, as you can see all the individual stats are bunched around the lower values and there is the total column as the indigo colored line.

Lets see the stats without the totals column, in a few different types of charts.

```python
gr_df.plot.bar(x='Name',y=['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed'],fontsize=15,figsize=(25,4), title='Grass Type Pokemon All Stats');
```

As you can see there are 66 values and 6 categories so the bar chart in its current form does not make for good analysis. Lets try it in subplots.

```python
gr_df.plot.bar(x='Name',y=['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed'],fontsize=15,figsize=(25,25), title='Grass Type Pokemon All Stats',subplots=True,sharex=True);
```

Well the colors look pleasing but we need something short and concise. How about a stacked bar chart?

```python
gr_df.plot.bar(x='Name',y=['HP','Attack','Defense','Sp. Atk','Sp. Def','Speed'],stacked=True, fontsize=15,figsize=(25,10), title='Grass Type Pokemon All Stats');
```

As we know pokemon are of three stages, so it would be a good idea to divide into the first, second and third stages. Lets try it.

```python
gr_df1=gr_df.iloc[::3,:]
gr_df1.plot.bar(x='Name',y=['HP'],figsize=(25,3), fontsize=15);
```

We have used iloc to pick out every 3rd pokemon starting from the first one, some pokemon have only two stages of evolution so we will have some errors in the stages of pokemon.

```python
gr_df.plot.bar(x='Name',y=['HP'],figsize=(25,5),fontsize=15)
a=[*range(30,121,30)]
pll.yticks(a);
```

Seperating the level 1 pokemon. We need to filter out pokemon with HP 30-55

From the complete grass type visualization sorted by HP, we can see roughly that the level 1 pokemon have an HP less than 50, or between 40 and 60. So we want to take out all these rows, and make a new data frame named as Level 1 Grass Type.

```python
gr_df1=gr_df.loc[(df['HP']<60)]
gr_df1.plot.bar(x='Name',y=['HP'],figsize=(10,3))
b=[*range(0,61,30)]
pll.yticks(b)
pll.rcParams['xtick.labelsize']=5
pll.rcParams['ytick.labelsize']=5
```

Now we can check the graphs for all other visualizations and check if there is something interesting.

```python
gr_df1.plot.bar(x='Name',y=['Attack','Sp. Atk'],figsize=(15,2.5))
a=[*range(0,101,20)]
pll.yticks(a)
gr_df1.plot.bar(x='Name',y=['Defense','Sp. Def'],figsize=(15,2.5))
a=[*range(0,101,20)]
pll.yticks(a)
pll.legend(loc='best');
```

```python
gr_df1.plot.bar(x='Name',y=['Attack','Defense','Speed'],figsize=(25,5), fontsize=20)
a=[*range(0,101,20)]
pll.yticks(a);
```

So this plot shows us a comparison of Attack, Defense and Speed stats of the level 1 pokemon in grass type.

Selecting some of the columns from a dataframe.

```python
grr=gr_df1[['Name','HP','Attack','Defense','Sp. Atk','Sp. Def','Speed']].copy()
grr.head()
grr.plot.bar(subplots=True, figsize=(25,10), layout=(3,2), sharey=True, fontsize=15)
a=[*range(0,121,20)]
pll.yticks(a)
x=grr['Name']
y=a
pll.xlabel('x')
pll.ylabel('y');
```
