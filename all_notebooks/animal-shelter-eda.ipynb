```python
Dogs_Sheltered = df['Animal Type'].value_counts()[0]
Cats_Sheltered = df['Animal Type'].value_counts()[1]
a = (Dogs_Sheltered/len(df))*100
b=(Cats_Sheltered/len(df))*100
print(f'There are {Dogs_Sheltered} dogs here')
print(f'There are {Cats_Sheltered} cats here')
print(f'There are {a}% of animals are dogs in the shelter')
print(f'There are {b}% of animals are cats in the shelter')
print(f'Cats and Dogs are Majority of the animals in the shelter standing at {a+b}% of total animals')

```

56% of the animals in the shelter are Dogs and 37% of the animals in the shelter are cats. Together they are standing at 94% of the total animals in the shelter.

```python
df['Date'] = pd.to_datetime(df.DateTime)
```

```python
df.info()
```

```python
df.drop(['DateTime'], axis = 1, inplace = True)
```

```python
df.info()
```

```python
df['IntakeYear'] = pd.DatetimeIndex(df.Date).year
df['IntakeMonth'] = pd.DatetimeIndex(df.Date).month
df['IntakeDay'] = pd.DatetimeIndex(df.Date).day
df['IntakeWeekDay'] = pd.DatetimeIndex(df.Date).weekday
```

```python
df['Intakehour'] = pd.DatetimeIndex(df.Date).hour
```

```python
df.drop(['MonthYear'], axis = 1, inplace = True)
```

```python
df.drop(['Found Location'], axis = 1, inplace = True)
```
