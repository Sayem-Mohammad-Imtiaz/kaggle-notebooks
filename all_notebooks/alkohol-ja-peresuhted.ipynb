**Alkoholi tarbimise ja peresuhetest seos.**
Uurimus viidi läbi kahes Portugali koolis. Õpilastel paluti hinnata 1-5 punkti süsteemis oma alkoholi tarbimisharjumusi, tervist, peresuhteid.

**Alkohol ja vanus**
Kõigepealt on tulpdiagramm, mis vaatleb alkoholi tarbimist vanuse kohta. 

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
df = pd.read_csv("../input/student-mat.csv")
df["Pstatus"] = df["Pstatus"].map({"A": "Lahku", "T": "Koos"})
df["famsize"] = df["famsize"].map({"GT3":"Rohkem kui 3", "LE3":"Vähem kui 3"})
```

```python
df1 = pd.DataFrame(df.groupby("age").aggregate({"Dalc":["mean"],"Walc":["mean"]}))
df1.plot.bar()
```

Lisaks sellele, et vanemaks saades tarbivad õpilased rohkem alkoholi, võib näha ka, et nooremad õpilased tarbivad tööpäevadel vähem kui nädalavahetustel. Vanemaks saades, see vahe ühtulustub.

```python
df2 = pd.DataFrame(df.groupby("famsize").aggregate({"Dalc": ["mean"], "Walc": ["mean"]})
                        .rename(columns={"Dalc":"Nädalapäevas alk", 
                                         "Walc":"Nädalavahetusel alk"}).round(2))
df2.plot.bar()
```

Peredes, kus on kolm või vähem last tarbitakse vähem, kui nelja ja rohkemate lastega peredes.

```python
df3 =  pd.DataFrame(df.groupby("age").aggregate({"Dalc": ["mean"], 
                                                    "Walc": ["mean"], 
                                                    "health":["mean"]})
                        .rename(columns={"Dalc":"Nädalapäevas alk", 
                                         "Walc":"Nädalavahetusel alk", 
                                         "health": "Tervis"}).round(2))
df3.plot()
```

**Peresuhted ja alkohol**
Küsitlusalustel paluti ka hinnata, kui kindlad on nende suhted pereliikmetega.

```python
df_sc1 = df.plot.scatter("famrel","Dalc",s=200,alpha=0.05)
df_sc2 = df.plot.scatter("famrel","Walc",s=200, alpha=0.05)
```

Hajuvusdiagrammidelt võib järeldada, et isegi kui peresuhted on halvad alkoholi tarbimine ei suurene.

**Kokkuvõetav tabel**

```python
df_lõpp = pd.DataFrame(df.groupby(["age","famsize","Pstatus"])["Dalc","Walc","health","famrel"].mean()
                       .round(2).rename(columns={"Dalc":"Nädalapäevas alk","Walc":"Nädalavahetusel alk",
                                      "health":"Tervis", "famrel":"Peresuhted"}))
df_lõpp.index.names = "Vanus", "Peresuurus", "Vanemate kooselavus"
df_lõpp
```

Kokkuvõtteks võib järeldada, et rohkem tarbivad suuremate peredega ja kus vanemad ei ela koos. Vanemad õpilased paraku joovad palju, isegi siis kui peresuhted on head, äkki probleem on kuskil mujal.
