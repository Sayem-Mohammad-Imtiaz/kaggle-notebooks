```python
#Modules for EDA
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')

%matplotlib inline
```

# **Loading the data.**

```python
df = pd.read_csv('../input/petrol-price-in-india/PetrolPriceIndia.csv')
df.shape
```

# **Dataset Info**

```python
df.info()
```

```python
pd.set_option('max_rows',36)
df
```

# **Setting 'Month-Year' as Index**

```python
df.set_index('Month-Year',inplace=True)
df
```

# **Analysis on Opening, Closing, Lowest and Highest of each year.**

```python
for year in range(18,21):#2018-2021
    plt.figure(figsize=(20,10))
    sns.lineplot(data=df.loc[f'Jan-{year}':f'Dec-{year}'])
    plt.title(f'Petrol prices in {2000+year}')
    plt.show()
    print('\n')
```

# **Year Wise Comparision on each column**

```python
for column in df.columns:
    start=0
    end = 12
    
    plt.figure(figsize=(20,10))
    for year in range(18,21):#2018-2021
        plt.plot(range(12),df[column][start:end],linestyle='dashed',label=f'{2000+year}-{column}')
        
        start += 12
        end += 12
    plt.legend()
    plt.show()
    
    print('\n')
```
