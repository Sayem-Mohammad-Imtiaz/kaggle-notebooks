```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
import itertools
class AprioriAlgorithm(object):
    def fit(self,X):
        
        self.X         = X 
        items          = []
        self.min_threshold_set = {}
        for i in self.X.keys():
            for j in self.X[i]:
                if (j in items)==False:
                    items.append(j)
                    
        self.items = items
        
    def Issublist(self,l1, l2):
        
        result = True
        
        for i in l1:
            if (i in l2)==False:
                result = False
                
        return result
    
    def Support(self, min_threshold, combinations=2):
        support   = {}

        for i in range(1, combinations+1):
            for subset in itertools.combinations(self.items, i):
                count = 0
                for j in self.X.keys():
                    if self.Issublist(list(subset), self.X[j])==True:
                        count += 1
                name = ''
                for x in range(len(list(subset))):
                    if x==0:
                        name = name + list(subset)[x]
                    else:
                        name = name + ", " + list(subset)[x]
                if round(count / len(self.X.keys()),2) >= min_threshold:
                    support[name] = round(count / len(self.X.keys()),2) 
                    
        print("Processing {} combinations | Sampling itemset size {} | Minimum Threshold {}".format(combinations, 
                                                                                                    len(self.items),
                                                                                                    min_threshold))
        plt.figure(figsize=(15,5))
        plt.bar(x=support.keys(),height=support.values(), color="teal", edgecolor="k")
        plt.xticks(rotation=90)
        plt.title("Apriori Result", fontsize=20)
        plt.xlabel("Items")
        plt.ylabel("Support")
        plt.grid()
        plt.show()

        return support
```
