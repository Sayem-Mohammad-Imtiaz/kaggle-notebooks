```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
titanic_df = pd.read_csv('/kaggle/input/cleaned-titanic-data-set-for-eda/titanic_cleaned.csv')
titanic_df.head()
```

```python
total_saved = titanic_df["Survived"].sum()
total_passenger= len(titanic_df["Survived"])
```

```python
saved_percentage= (total_saved/total_passenger)*100
print (f"{saved_percentage} % of the total passenger was saved in Titanic")
```

```python
women_saved = titanic_df.loc[titanic_df["Sex"]== "female"]["Survived"]
```

```python
women_saved_percentage = (women_saved.sum()/women_saved.count())*100
print(f"{women_saved_percentage}% of the total Women was saved in Titanic")
```

```python
men_saved = titanic_df.loc[titanic_df["Sex"]== "male"]["Survived"]
```

```python
men_saved_percentage = (men_saved.sum()/men_saved.count())*100
print(f"{men_saved_percentage}% of the total Men was saved in Titanic")
```

### Majority of the people were between an age of 18 to 34

```python
plt.figure(figsize=(12, 6))
plt.title("Distribution of Age in Titanic")
plt.xlabel('Age')
plt.ylabel('Number Passengers')
plt.hist(titanic_df["Age"], bins=np.arange(10,80,5), color='purple')
```

### We can clearly see that Gender played an important role at a survival because women have been prioritize in rescue operation 

```python
sns.set_style("whitegrid")
sns.countplot(x= "Survived", hue= "Sex", data= titanic_df)
```

### We can clearly see that Pclass played an very important role in survival rate, Women Passenger class 1, 2, and 3 had a survival chance of 95%, 90% and 50% Respectively. Whereas Male Passenge class of 1, 2 and 3 had a survival chance of 39%, 15% and 13% Respectively

```python
plt.figure(figsize=(12,6))
sns.set_style("whitegrid")
sns.catplot(x= "Pclass", y="Survived",hue= "Sex",data = titanic_df, kind = "bar")
```

```python
sns.set_style("whitegrid")
sns.countplot(x= "Survived", hue= "Pclass",data =titanic_df)
```
