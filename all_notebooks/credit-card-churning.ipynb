# Importing Library

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("/kaggle/input/credit-card-customers/BankChurners.csv")
df.head()
```

```python
df.head()
```

# Dropping first and last two column

```python
df.drop(['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2','CLIENTNUM'],axis=1,inplace=True)
```

```python
df.head()
```

# Checking for null values

```python
df.isnull().sum()
```

# Mapping Attrition_Flag

```python
df['Attrition_Flag'] = df['Attrition_Flag'].map({'Existing Customer': 1, 'Attrited Customer': 0})
```

# Correlation

```python
sns.heatmap(df.corr())
```

# Removing Unwanted columns

```python
df = df.drop(columns = 'Avg_Open_To_Buy')
df = df.drop(columns = 'Credit_Limit')
```

# Balance / Imbalance

```python
df['Attrition_Flag'].value_counts()
```

# One Hot Encoding

```python
df = pd.get_dummies(df,drop_first = True)
```

# Dependent and Independent Variable

```python
x = df.drop(columns = 'Attrition_Flag')
y = df['Attrition_Flag']
```

# Test set and Train set

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators = 10)

rf.fit(x_train,y_train)
```

```python
y_pred = rf.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score

accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

```python
precision_score(y_pred,y_test)
```

```python
sns.distplot(y_test-y_pred)
```
