```python
df_forest_fire = pd.read_csv('../input/forest-fires-in-brazil/amazon.csv', encoding='ISO-8859-1')
```

# Adding up all the fires for years 

```python
df_forest_fire_by_year = df_forest_fire.groupby(['year']).number.agg(sum)
df_forest_fire_by_year =pd.DataFrame(df_forest_fire_by_year)
df_forest_fire_by_year.number = df_forest_fire_by_year.number.map(lambda p: round(p))
df_forest_fire_by_year.plot(ylabel='Total fires by year', figsize=(10,5))
df_forest_fire_by_year.T
```

# Searching total fires per year each state

```python
df_forest_fire_by_year = df_forest_fire.groupby(['year','state']).number.agg(sum)
df_forest_fire_by_year = pd.DataFrame(df_forest_fire_by_year).reset_index()
df_forest_fire_by_year.number = df_forest_fire_by_year.number.map(lambda p: round(p))
year_wanted = int(input('Enter the desired year between 1998 and 2017: '))
df_year_wanted = df_forest_fire_by_year.loc[df_forest_fire_by_year.year == year_wanted]
df_year_wanted.sort_values(by='number', ascending=False).reset_index(drop=True)
```

# Searching fires by state and history of fires

```python
state_search = df_forest_fire.groupby(['state','year']).number.agg(sum).reset_index()
state_search.number = state_search.number.map(lambda p: round(p))
state_wanted = input('Enter the state name with the first letter capitalized: ')
df_state_search = state_search.loc[state_search.state == state_wanted ].set_index('year',drop=True)
df_state_search.plot(figsize=(10,5), ylabel='number of fires')
df_state_search.reset_index()
```
