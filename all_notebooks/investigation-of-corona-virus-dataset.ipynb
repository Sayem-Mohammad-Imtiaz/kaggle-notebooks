```python
Turkey=df.country=="Turkey" #First confirmed in 11.03.2020
Turks=df[Turkey]
Turks
```

```python
Turks.loc[:,["deaths","recovered"]]
```

```python
filter_2=Turks.deaths>=1   # After 6 days later from first confirmed there occured first death on 17th of March.
de_rec=Turks[filter_2]
de_rec
```

```python
filter_3=Turks.recovered>=1
de_rec2=Turks[filter_2&filter_3]  #Difference between deaths and recovered can be seen clearly seen day by day. By time,number of recovered increases much more than deaths.
de_rec2
```

```python
avdpday=de_rec.deaths[116218]/len(de_rec.serial_number) # Mean of deaths per day after first death has been seen in Turkey.
print(avdpday)
```

```python
de_rec.serial_number=[i for i in range(1,len(de_rec.serial_number)+1)]
print("Old Datas about Turkey: \n",de_rec)
daily_deaths1=[index for index,value in de_rec.deaths.items()] 
daily_deaths2=[value for index,value in de_rec.deaths.items()]#Starting from this row, it was tried to acquire daily deaths 
                                                              #because there exist cummulative deaths.
daily_deaths3=daily_deaths2.copy()
daily_deaths3.pop(len(daily_deaths3)-1)
daily_deaths3.insert(0,0)
daily_deaths3=np.array(daily_deaths3)
daily_deaths2=np.array(daily_deaths2)
daily_deaths7=daily_deaths2-daily_deaths3
daily_deaths7=list(daily_deaths7)
de_rec["daily_deaths"]=daily_deaths7
de_rec["death_rate"]=["low" if each<=avdpday else "high" for each in de_rec.daily_deaths]
print("New Datas about Turkey: \n",de_rec)
print(de_rec.daily_deaths.mean())
print(np.mean(de_rec.daily_deaths))
```
