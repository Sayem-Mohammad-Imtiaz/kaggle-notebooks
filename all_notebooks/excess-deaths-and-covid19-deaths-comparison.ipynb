```python
import numpy as np
import pandas as pd
import datetime as datetime
import matplotlib.dates as dates
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import os
```

```python
data= pd.read_csv("../input/covid19-deaths-dataset/all_weekly_excess_deaths.csv")
data[0:2]
```

```python
data['region'].unique()
```

```python
data_usa=data[data['region']=='United States']
data_usa[0:2]
```

```python
data_uk=data[data['region']=='Britain']
data_uk[0:2]
```

```python
#1
fig=make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=data_usa['start_date'],y=data_usa['total_deaths'],name="total_deaths"),secondary_y=False,)
fig.add_trace(go.Scatter(x=data_usa['start_date'],y=data_usa['covid_deaths'],name="covid_deaths"),secondary_y=False,)

fig.update_layout(autosize=False,width=700,height=500,title_text="Deaths in USA")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Number",secondary_y=False)
fig.show()
```

```python
#2
fig=make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=data_usa['start_date'],y=data_usa['expected_deaths'],name="expected_deaths"),secondary_y=False,)
fig.add_trace(go.Scatter(x=data_usa['start_date'],y=data_usa['total_deaths'],name="total_deaths"),secondary_y=False,)
fig.add_trace(go.Scatter(x=data_usa['start_date'],y=data_usa['excess_deaths'],name="excess_deaths"),secondary_y=False,)
fig.add_trace(go.Scatter(x=data_usa['start_date'],y=data_usa['covid_deaths'],name="covid_deaths"),secondary_y=False,)

fig.update_layout(autosize=False,width=700,height=500,title_text="Excess Deaths in USA")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Number",secondary_y=False)
fig.show()
```

### Excess_deaths is almost equal to real covid_deaths.

```python
#3
fig=make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=data_usa['start_date'],y=data_usa['covid_deaths_per_100k'],name="covid_deaths_per_100k in USA"),secondary_y=False,)
fig.add_trace(go.Scatter(x=data_uk['start_date'],y=data_uk['covid_deaths_per_100k'],name="covid_deaths_per_100k in UK"),secondary_y=False,)

fig.update_layout(autosize=False,width=700,height=500,title_text="Covid Deaths per 100k people in USA and UK")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Number",secondary_y=False)
fig.show()
```
