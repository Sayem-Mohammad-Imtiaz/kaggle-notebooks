```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/kaggle/input/voicegender/voice.csv')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data=LabelEncoder()
```

```python
df['label']=data.fit_transform(df['label'])
```

```python
df
```

```python
sns.relplot(x='meanfreq',y='meanfun',hue='label',kind='scatter',data=df)
```

```python
x=df.drop(['label','centroid'],axis='columns')
```

```python
y=df['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict(x_test)
```

```python
a=x_test['result']=model.predict(x_test)
```

```python
x_test
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,a)
```

```python
cm
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(cm,annot=True)
plt.xlabel('actual')
plt.ylabel('y_predict')
```
