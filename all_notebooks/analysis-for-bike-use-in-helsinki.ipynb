```python
def distance(ex):
    if ex <= 1000:
        return 0
    elif 1000 < ex <=2000:
        return 1
    elif 2000 < ex <=3000:
        return 2
    elif 3000 < ex <=4000:
        return 3
    elif 4000 < ex <= 5000:
        return 4
    else:
        return 5
```

```python
df['dis_lebel']=df['distance (m)'].apply(distance)
```

```python
df.groupby(['dis_lebel'])['dis_lebel'].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
#94% of bike use is within 5km. 25% of bike use is withn 1km. More than 75% of bike use is withn 3km.
```

```python
df1=df.groupby(['departure_name','return_name'])['distance (m)'].agg(['count','mean'])
```

```python
df1
```

```python
#top20 roots analysis
```

```python
df_20=pd.DataFrame(df1.sort_values(by='count',ascending=False).head(20))
```

```python
df_20
```

```python
df_20=df_20.rename(columns={'mean':'mean_distance'})
```
