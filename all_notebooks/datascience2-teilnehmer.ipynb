# Hausaufgabe

Schreibe ein Programm, dass dir eine Liste erstellt mit 100 zufälligen Zahlen (zwischen 1 und 5).

Dafür brauchst du
* Eine leere Liste
* Eine **for**-Schleife
* Die Funktion **randint()** von der Bibliothek **random**
* Die Methode **append()** um die Werte in die Liste zu schreiben

```python
# Lösung 1
from random import randint

zufallszahlen = []
for i in range(100): 
    zufallszahlen.append(randint(1,5))
    
print(zufallszahlen)
```

```python
# Lösung 2

# List-Comprehension
zufallszahlen = [randint(1,5) for i in range(100)]

print(zufallszahlen)
```

# Zusatzaufgabe

Bestimme den Mittelwert deiner Liste

## arithmetisches Mittel

*$\bar x = \frac{x_1 + x_2 + ... + x_n}{n}$*

```python
# Summe aller Werte der Liste:
summe = sum(zufallszahlen)
print(summe)

# Die Länge der Liste:
laenge = len(zufallszahlen)
print(laenge)

# Das arithmetische Mittel:
mean = summe / laenge
print(mean)
```

## Median

Der Median ist der Zentralwert einer sortierten Liste

* Sortiere die Liste
* Bestimme die Länge
* Ist die Anzahl der Elemente der Liste ungerade so bildet der Wert an Position [Länge // 2] den Zentralwert
* Ist die Anzahl gerade so liegt der Zentralwert zwischen den Werten an den Positionen [Länge // 2 -1] und [Länge // 2]. Aus den Werten an diesen Positionen wird der Mittelwert gebildet

```python
liste = [randint(1,5) for i in range(100)]

liste.sort()
laenge = len(liste)
if laenge % 2 == 0 and laenge != 0:
    median = (liste[laenge // 2 - 1] + liste[laenge // 2]) / 2
elif laenge % 2 == 1:
    median = liste[laenge // 2]
    
print(median)
```

# Numpy

Ehe man an anfängt das Rad neu zu erfinden, kann man auch gleich auf Programme zurückgreifen, die eine Lösung parat haben. Numpy ist eine Bibliothek, die eine für uns sehr nützliche Datenstruktur bereitstellt: der **array**.

```python
# importiere Numpy
import numpy as np
```

```python
# Unsere Liste mit unterschiedlichen Datentypen können wir an eine Funktion in Numpy weitergeben:
liste3 = [1, 2, "3", 4]

# np.array wandelt die Liste in einen Array um. Dabei können wir beim erstellen des Arrays definieren,
# welchen Datentyp die Elemente haben sollen (hier z.B. "int" für Integer)
array = np.array(liste3, dtype = "int")

print(array)
```

Ebenfalls sehr komfortable an Numpy: Es vereinfacht die mathematische Verarbeitung von numerischen Wwerten in Datenstrukturen.

So lassen sich mit den entsprechenden Funktionen z.B.:
* Mittelwert (mean)
* Standardabweichung (std)
* oder der Median (median)

bestimmen

```python
array = np.array([1,2,3,4,99])

print("Mittelwert: ", np.mean(array))
print("Median: ", np.median(array))
print("Standardabweichung", np.std(array))
```

Arrays lassen sich auch mit anderen Arrays verrechnen (eventuell ist hier wichtig auf die Form der Arrays zu achten)

```python
# Erzeuge zwei Arrays mit 1000 Zufallszahen zwischen 1 und 5
zfz1 = np.random.randint(1,5, size = 1000)
zfz2 = np.random.randint(1,5, size = 1000)

# Addiere beide Arrays (Analog zur Addition zweier Vektoren)
zfz_ges = zfz1 + zfz2
```

```python
# Kleiner Spoiler:
# Um das ganze zu Visualisieren
import matplotlib.pyplot as plt

plt.hist(zfz_ges, bins=7)
plt.show()
```

```python
# Wir erkennen eine sogenannte Normalverteilung (erkennbar an der Glockenform)
# Das spiegelt sich auch in den statistischen Werten wieder

# Mean bildet ungefähr den Bereich ab, der den höchsten Punkt der Glocke abbildet
print(np.mean(zfz_ges))

# Die Standardabweicung verrät, wie schmal oder breit die Glocke ist.
# Zudem hilft sie uns, eine Vorhersage für weitere Daten zu geben.
print(np.std(zfz_ges))
```

```python
# Im Vergleich eine uniforme Verteilung (Jeder Wert hat die gleiche Wahrscheinlichkeit aufzutreten)
plt.hist(zfz1, bins = 4)
plt.show()
```
