### Loading the Data

### Summary of the Data

```python
nyc_data.columns
```

```python
features = ['id', 'name', 'host_id', 'host_name', 'neighbourhood_group_cleansed', 'neighbourhood_cleansed', 'latitude', 'longitude', 'room_type', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365']
nyc_data_filtered = nyc_data[features]
```

```python
# Missing Values
nyc_data_filtered.isnull().any()
```

```python
nyc_data_filtered['reviews_per_month'].fillna(0, inplace=True)
```

### Exploring and Visualizing

Let's explore and visualize our filtered data, so we can better understand the correlation between the features.

- #### Neighbourhood Group Cleansed

```python
sizes = nyc_data_filtered['neighbourhood_group_cleansed'].value_counts()
labels = nyc_data_filtered['neighbourhood_group_cleansed'].value_counts().index
plt.figure(figsize=(16,9))
plt.title('Neighbourhood Group Cleansed')
plt.pie(x=sizes, labels=labels, shadow=True, autopct='%1.1f%%')
plt.show()
```

As we can see from the pie chart above, Airbnb listings in New York are close to Manhattan.

```python
x = nyc_data_filtered['longitude']
y = nyc_data_filtered['latitude']
hue = nyc_data_filtered['neighbourhood_group_cleansed']
plt.figure(figsize=(16,9))
plt.title('Map of the Neighbourhood Group Cleansed')
sns.scatterplot(x=x, y=y,hue=hue)
plt.show()
```

- #### Room Type

Let's see the type of room.

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(16,8))
plt.title("Room Type")
sns.countplot(x='room_type', data=nyc_data_filtered, palette="muted")
plt.show()
```

We can see that the entire house/apt has the highest share, followed by the private room, so we have the shared room as the penultimate shared room and for the Hotel room as the least preferred one.

Let's see now, the room type for a neighbourhood group.


```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(16,8))
plt.title("Room Type on Neighbourhood Group")
sns.countplot(x='neighbourhood_group_cleansed', hue='room_type', data=nyc_data_filtered, palette="muted")
plt.show()
```

The bar plot shows that the Entire home/apt is most listed near Manhattan, while Entire home/apt and Private room near Brooklyn are almost equally distributed.
