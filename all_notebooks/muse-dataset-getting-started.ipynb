```python
from IPython.display import Audio
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import umap
import umap.plot
from bokeh.plotting import show, save, output_notebook, output_file
from bokeh.resources import INLINE
output_notebook(resources=INLINE)
```

```python
df = pd.read_csv('/kaggle/input/muse-the-musical-sentiment-dataset/muse_v3.csv').astype({'valence_tags': np.float32, 'arousal_tags':np.float32, 'dominance_tags':np.float32})
```

```python
df[df['artist']=='Leonard Cohen']['genre'].value_counts()
```

```python
subsample_df = df[(df['number_of_emotion_tags'] > 6) & ~df['genre'].isna()].reset_index(drop=True)
embeddings = subsample_df[['valence_tags', 'arousal_tags', 'dominance_tags']].values
kmeans = KMeans(n_clusters=3, random_state=42).fit(embeddings)
mapper = umap.UMAP(n_neighbors=100).fit(embeddings)
f = umap.plot.interactive(mapper, point_size=3, labels=kmeans.labels_, hover_data=subsample_df[['genre']], width=600, height=600, background='black')
show(f)
```
