# 2. Clustering

```python
from sklearn import preprocessing
from sklearn.cluster import KMeans
```

```python
X=df.drop(['State','code'],axis=1)
sc=preprocessing.StandardScaler()
sc.fit(X)
X_norm=sc.transform(X)
print(type(X_norm))
```

```python
cls = KMeans(n_clusters=4)
cluster = cls.fit_predict(X_norm)
```

```python
cluster=pd.DataFrame(cluster)
```

1) Mapping by Clusters

```python
df['cluster']=cluster
df.head()
```

```python
fig = px.choropleth(df,locationmode='USA-states',locations='code',color='cluster',scope='usa',color_continuous_scale=px.colors.sequential.GnBu,
                 range_color=(0,3),title='Mappingg by Clusters',height=500
    )
fig.show()
```

# 3.Premium Prediction Model

1) Linear Regression by Featuress

```python
from sklearn.linear_model import Ridge,Lasso,LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
from sklearn.model_selection import KFold
```

```python
x1=df.drop(['State','code','Ins_Premium','cluster'],axis=1)
y=df['Ins_Premium']
```

```python
x_train1,x_test1,y_train,y_test=train_test_split(x1, y,train_size=0.8,random_state=123)
```

```python
model=LinearRegression()
```

```python
scores=[]
kf=KFold(n_splits=4,shuffle=True,random_state=71)
for tr_idx,va_idx in kf.split(x_train1):
    tr_x,va_x=x_train1.iloc[tr_idx],x_train1.iloc[va_idx]
    tr_y,va_y=y_train.iloc[tr_idx],y_train.iloc[va_idx]
    
    model.fit(tr_x,tr_y)
    va_pred1=model.predict(va_x)
    score_rmse1=np.sqrt(mean_squared_error(va_y,va_pred1)).mean()
    score_mae1=mean_absolute_error(va_y,va_pred1).mean()
    score_r21=r2_score(va_y,va_pred1).mean()
```

```python
print('rmse1:',score_rmse1)
print('mae1:',score_mae1)
print('R21:',score_r21)
```
