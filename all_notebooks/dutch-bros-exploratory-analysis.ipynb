```python
df = pd.read_csv('/kaggle/input/dutch-bros-locations/dutch_bros_locations.csv')
df.head()
```

```python
df.info()
```

```python
by_state = df.groupby('state')['city'].count().sort_values(ascending=False)
by_state
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax = sns.barplot(x=by_state.index, y=by_state)
ax.set_title('Count of Dutch Bros Locations by State')
ax.set_xlabel('State')
ax.set_ylabel('Count of Locations')

plt.show()
```
