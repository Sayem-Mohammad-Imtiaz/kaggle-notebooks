```python
data = pd.read_csv("../input/automoviles/automoviles.csv", sep=";") 
```

```python
data = data.drop(columns=["normalized-losses","tipo_motor", "marca","tipo_combustible","tipo_turbo","estilo","drive-wheels","ubicacion_motor","sistema_combustible"]) 
```

```python
data["num_puertas"].value_counts()
```

```python
data = data.fillna({"num_puertas": "four"})
```

```python
data["num_cilindros"].value_counts()
```

```python
new_values = {"num_puertas":     {"four": 4, "two": 2},
                "num_cilindros": {"four": 4, "six": 6, "five": 5, "eight": 8,
                                  "two": 2, "twelve": 12, "three":3 }}
```

```python
data = data.replace(new_values)
data.head()
```

```python
data.dtypes
```

```python
data.head(40)
```

```python
X = data.drop("precio", axis=1)
y = data["precio"]
```

```python
X_train, X_test, Y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape, Y_train.shape, X_test.shape, y_test.shape
```

```python
#X_test = X_test.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, Y_train)
Y_pred = logreg.predict(X_test)
acc_log = round(logreg.score(X_train, Y_train) * 100, 2)
acc_log
```

```python
print(Y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
random_forest = RandomForestRegressor(n_estimators=10)
random_forest.fit(X_train, Y_train)
Y_pred = random_forest.predict(X_test)
random_forest.score(X_train, Y_train)
acc_random_forest = round(random_forest.score(X_train, Y_train) * 100, 2)
acc_random_forest
```
