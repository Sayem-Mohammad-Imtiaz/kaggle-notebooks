```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
life_exp = pd.read_csv("../input/countries-life-expectancy/Life expectancy.csv")
life_exp.head()
```

```python
life_exp.tail()
```

```python
#Animation
import plotly.express as px
fig = px.bar(life_exp, x="Entity", y="Life expectancy", color="Entity",
  animation_frame="Year", animation_group="Entity", range_y=[0,100])
fig.show();
```

```python
#Line chart 
import plotly.express as px
fig = px.line(life_exp, x="Year", y="Life expectancy", color="Entity",
              line_group="Entity", hover_name="Entity")
fig.show()
```

```python
import plotly.express as px
fig = px.area(life_exp, x = "Year", y= "Life expectancy", color = "Entity")
fig.show()
```

```python
fig = px.scatter(life_exp, x = "Year", y = "Life expectancy", animation_frame = "Year", animation_group = "Entity", size= "Life expectancy", color = "Entity", hover_name = "Entity", log_x = True, range_x = [1800, 2016], range_y = [20, 100])
fig.show()
```

```python
Australia = life_exp[life_exp['Entity']== 'Australia']
Australia.head()
```

```python
#Years with higher life expectancy rates
Australia.nlargest(5, 'Life expectancy')
```

```python
g = sns.relplot(x="Year", y="Life expectancy", kind="line", data=Australia)
g.fig.autofmt_xdate()
```
