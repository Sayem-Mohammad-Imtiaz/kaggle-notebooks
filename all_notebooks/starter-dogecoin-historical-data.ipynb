This dataset contains one file `DOGE-USD.csv`.

```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv("../input/dogecoin-historical-data/DOGE-USD.csv")
```

The `Date` column is a string that should be parsed as datetime datatype.

```python
df.Date = pd.to_datetime(df.Date)
```

```python
df.dtypes
```

```python
df
```

Let us plot a candlestick chart

```python
t = df.tail(n=50)
```

```python
fig = go.Figure(
    data=[
        go.Candlestick(
            x=t["Date"],
            open=t["Open"],
            high=t["High"],
            low=t["Low"],
            close=t["Close"],
        )
    ]
)
```

```python
fig.show()
```
