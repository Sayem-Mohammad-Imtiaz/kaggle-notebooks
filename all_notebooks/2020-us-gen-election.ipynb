```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../input/2020-us-general-election-turnout-rates/2020 November General Election - Turnout Rates.csv')
```

```python
df.head()
```

```python
print('There are {} rows and {} columns in our dataset.'.format(df.shape[0], df.shape[-1]))
```

```python
df.info()
```

```python
percent = (df.isnull().sum()*100)/len(df)
percent.sort_values(ascending=False)
```

```python
colm = ['State Abv', 'Source','Vote for Highest Office (President)']
df.drop(colm, axis=1, inplace=True)
```

```python
df.columns
```

```python
df.drop(index=[0], inplace=True)
```

```python
df.head()
```

```python
count =['Total Ballots Counted (Estimate)', 'Voting-Eligible Population (VEP)', 'Voting-Age Population (VAP)',
       'Prison', 'Probation', 'Parole', 'Total Ineligible Felon']

perc = ['VEP Turnout Rate', '% Non-citizen']

for val in count:
    df[val]=df[val].str.replace(',','')

for val in perc:
    df[val]=df[val].str.replace('%', '')

df[count] = df[count].astype('int')
df[perc] = df[perc].astype('float')
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(), annot=True)
```

```python
df.head()
```

```python
df1 = df.set_index('State')
plt.figure(figsize=(12,6))
sns.barplot(x='State', y='VEP Turnout Rate', data=df, order=df1['VEP Turnout Rate'].sort_values(ascending=False).index[:15])
plt.xticks(rotation=90)
plt.xlabel('State', weight='bold')
plt.ylabel('VEP Turnout Rate (%)', weight='bold')
plt.show()
```

```python
df1 = df.set_index('State')
plt.figure(figsize=(12,6))
sns.barplot(x='State', y='Voting-Age Population (VAP)', data=df,
           order=df1['Voting-Age Population (VAP)'].sort_values(ascending=False).index[:15] , color='y')
plt.xticks(rotation=90)
plt.xlabel('State', fontsize=25)
plt.ylabel('Voting Age Population', weight='bold', fontsize=25)
plt.show()
```

```python
df_temp = df.set_index('State')
fig, ax = plt.subplots(2, 2, figsize=(16,16))
for col, ax in zip(['Prison','Probation','Parole','Total Ineligible Felon'], ax.flat):
    dict_ = df_temp[col].sort_values(ascending=False).head(10).to_dict()
    ax.pie(x=list(dict_.values()), labels=dict_.keys(), shadow=True, startangle=90)
    ax.set_title(col, weight='bold')
plt.show()
```

```python
df_temp = df.set_index('State')
df_temp = df_temp.sort_values('Voting-Eligible Population (VEP)', ascending=False).head(10)
sns.set(style='darkgrid')
df_temp[['Prison','Probation','Parole','Total Ineligible Felon']].plot(kind="bar", figsize=(14,8), width=0.7)
plt.xticks(rotation=90)
plt.xlabel('State', weight='bold')
plt.ylabel('Population', weight='bold')
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)
plt.legend(shadow=True, facecolor='lightyellow')
plt.show()
```
