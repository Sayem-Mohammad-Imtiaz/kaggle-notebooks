# Visualization

```python
#We create box plot in feature tempo for every single genre.
px.box(data, x='genre',y='tempo')
```

Song of Dark trap, Under ground Rap, Trap Metal, Emo, Rap, RnB, Pop, Hip Hop has taller range Q1 and Q3 from median different to another. 

#Let see in 3D for The Loudness, Danceability,Tempo

```python
px.scatter_3d(data, x='loudness', y='danceability', z='tempo', color='duration_ms', color_continuous_scale='cividis', opacity=0.75)
```

It was too crowded to see. Let's move to line, and change by genre

```python
px.line_3d(data,x='tempo', y='loudness', z='genre', color='genre')
```

We can see taht there're two categories that has short tempo such hardstyle, trap, trance, techhouse, and long tempo is the rest.
