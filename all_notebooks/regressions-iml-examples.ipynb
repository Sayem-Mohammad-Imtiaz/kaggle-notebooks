# Dataset formatting/cleaning

```python
df = pd.read_csv("../input/airbnb-price-prediction/train.csv")
df.head(10)
```

```python
filtered = df[["log_price", "accommodates", "bathrooms", "number_of_reviews", "review_scores_rating", "bedrooms", "beds"]]
filtered.head()
```

# Simple Linear

```python
y = filtered[["log_price"]]
y.head()
```

```python
x1 = filtered[["accommodates"]]
x1.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x1, y, test_size=0.33, random_state=42)
```

```python
model = lm.LinearRegression()
results = model.fit(x_train,y_train)
print(model.intercept_, model.coef_)
```

```python
#Result: Scikit-Learn
yp2 = model.predict(x_test)
print(yp2)
```

```python
#Linear Regression representation using scatter plot
plt.scatter(x_train,y_train)
plt.plot(x_test,yp2, color="red")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(yp2, y_test, squared=False)
```

# Multiple Linear Regression

```python
inputDF = filtered[["accommodates", "bathrooms", "bedrooms", "beds"]]
x_train, x_test, y_train, y_test = train_test_split(inputDF, y, test_size=0.33, random_state=42)
```
