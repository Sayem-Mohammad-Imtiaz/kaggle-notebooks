```python
plt.figure(figsize = (20, 8))
sns.set(style = "darkgrid")
ax = sns.countplot(y = 'Job Title', data = df,
              order = df['Job Title'].value_counts().iloc[:10].index)
for p in ax.patches:
    ax.annotate(int(p.get_width()), ((p.get_x() + p.get_width()), p.get_y()), xytext = (-2, -20),
                fontsize = 16, color = '#004d00', textcoords = 'offset points', horizontalalignment = 'right')
plt.title('Top 10 Job Profile')
plt.show()
```

#### Here, I have plotted the top 10 job profiles for which jobs have been posted. We can see more Jobs have been posted for Sales Executive and Business Development Executive as compared to other Job Profiles which is further followed by PHP Developer and Business Development Manager.

```python
from wordcloud import WordCloud
```

```python
skills = df['Key Skills'].to_list()
skills = [str(s) for s in skills]
skills = [s.strip().lower()  for i in skills for s in i.split("|")]
skills = " ".join(w for w in skills)
wc = WordCloud(width = 2000, height = 1000).generate(skills)
plt.figure(figsize = (16, 8))
plt.imshow(wc)
plt.axis('off')
plt.show()
```

#### In above WordCloud we can see the most common skills required are:
* Business Development
* Sales Executive
* Technical Support
* Customer Service
* Software Developer
