```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
# Breast cancer dataset for classification
cancer = load_breast_cancer()

print('Keys:\n\n', cancer.keys())
print('\n--------------------------------------------------------------------------------------------------------------')
print ('\n\nFeature Names:\n\n', cancer.feature_names)
print('\n--------------------------------------------------------------------------------------------------------------')
print ('\n\nTarget Names:\n\n', cancer.target_names)
print('\n--------------------------------------------------------------------------------------------------------------')
print('\n\nData:', cancer.target)
print('\n--------------------------------------------------------------------------------------------------------------')
print('\n\nDESCR:', cancer.DESCR)
print('\n--------------------------------------------------------------------------------------------------------------')
print('\n\nfilename:', cancer.filename)
print('\n--------------------------------------------------------------------------------------------------------------')
```

```python
data = pd.DataFrame(cancer.data, columns=[cancer.feature_names])
data['Target'] = pd.Series(data=cancer.target, index=data.index)
```

<h1 style="background-color:LimeGreen; font-family:newtimeroman; font-size:180%; text-align:center; border-radius: 15px 50px;"> 2) Load and split example data </h1>

```python
import evalml

X, y = evalml.demos.load_breast_cancer()
X_train, X_test, y_train, y_test = evalml.preprocessing.split_data(X, y, problem_type='binary')
```

```python
from evalml.automl import AutoMLSearch
automl = AutoMLSearch(X_train=X_train, y_train=y_train, problem_type='binary')
automl.search()
```

<h1 style="background-color:LimeGreen; font-family:newtimeroman; font-size:180%; text-align:center; border-radius: 15px 50px;"> 3) View pipeline rankings </h1>

```python
automl.rankings
```

<h1 style="background-color:LimeGreen; font-family:newtimeroman; font-size:180%; text-align:center; border-radius: 15px 50px;"> 4) Get best pipeline and predict on new data </h1>

```python
pipeline = automl.best_pipeline
pipeline.fit(X_train, y_train)
prediction = pipeline.predict(X_test)
```

```python
data['y_test'] = prediction.to_series()
#data.to_csv('submission.csv',index=False)
data['y_test'].head()
```
