12 - quality (score between 0 and 10)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
#sns.set_style('darkgrid')

colors = ['#B90276','#50237F', '#005691', '#008ECF','#E20015', '#00A8B0', '#78BE20', '#006249', '#525F6B']

sns.set_palette(sns.color_palette(colors))
```

```python
data = pd.read_csv("/kaggle/input/red-wine-quality-cortez-et-al-2009/winequality-red.csv", sep = ",")
df = pd.DataFrame(data)
df.head()
```

```python
df.quality.unique()
```

```python
df.info()
```

```python
sns.countplot(df['quality'], data = df)
```

```python
i=1
plt.figure(figsize = (13,17))
for col in df.columns:
    plt.subplot(4,3,i)
    sns.histplot(df[col])
    #plt.title(col)
    i+=1
    
```

# Let's analyse how different parameters are affecting the qulity of wine

```python
i=1
plt.figure(figsize=(15,15))
for col in df.columns:
    plt.subplot(4,3,i)
    sns.boxplot(y=df[col], x = df['quality'])
    i+=1
```

***Good quality wine should have high alcoho between 11-13, low pH, average density, very low chlorides, high citric content***

# Plotting corelation 

```python
corr = df.corr()
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(15, 10))
    ax = sns.heatmap(corr, mask=mask, vmax=.3, square=True, cmap = 'coolwarm', linewidth = 0.5, annot = True)
```

# Let's analyse the data for wine quality > 6 score

```python
dfq = df[(df['quality']>=6)]
dfq.head()
```

```python
#plt.figure(figsize=(25,12))
sns.pairplot(dfq, vars = ['alcohol','volatile acidity', 'citric acid','sulphates', 'fixed acidity'], hue = 'quality', palette = 'deep')
```

```python
i=1
plt.figure(figsize=(15,15))
cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']
for col in cols:
    plt.subplot(4,3,i)
    sns.boxplot(y=dfq[col], x = dfq['quality'])
    i+=1
```
