# 1. Import Librairies

```python
import numpy as np 
import pandas as pd
from joblib import load
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

---

# 2. Loading the Dataset

```python
df = pd.read_csv("../input/cancermizu/df_test.csv")
df.drop(columns="Unnamed: 0", axis=1, inplace=True)
df.head()
```

```python
y = df["y"]
X = df.drop(columns="y", axis=1)

X.shape, y.shape
```

---

# 3. Loading the Model

```python
model = load("../input/cancermizu/model.joblib")
```

---

# 4. Performance on the Test Set

```python
# Predict
y_pred = model.predict(X)

# Confusion Matrix
cm = confusion_matrix(y, y_pred, normalize='true')

# Graph
plt.figure(figsize=(12, 7))
plt.title('Accuracy Score on the Test Set: ' + str(accuracy_score(y, y_pred).round(4)), size=25)
sns.heatmap(cm, annot=True, fmt='.2%', cmap='Blues')
plt.xlabel('Predicted Values', size=20)
plt.ylabel('True Values', size=20)
plt.show()
```
