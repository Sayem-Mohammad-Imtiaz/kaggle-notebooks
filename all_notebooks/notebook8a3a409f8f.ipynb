```python
x = new_data.iloc[:, : 34].values
y = new_data.iloc[:,-1].values
```

```python
#train test split
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
xtrain =sc.fit_transform(xtrain)
xtest =sc.transform(xtest)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=100,criterion='gini', n_jobs=-1,random_state=40,verbose=3)
rfc.fit(xtrain,ytrain)
```

```python

y_pred=rfc.predict(xtest)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
cm = confusion_matrix(ytest,y_pred)
cm
```

```python
cr = classification_report(ytest,y_pred)
print(cr)
```

```python
score = accuracy_score(ytest,y_pred)
score
```

```python
#Decison Tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='entropy',
    max_depth=50,
    min_samples_split=100,
    random_state=100)
dt.fit(xtrain,ytrain)
```

```python
y_pred1=dt.predict(xtest)
y_pred1
```

```python
cm1 = confusion_matrix(ytest,y_pred1)
cm1
```

```python
cr1 = classification_report(ytest,y_pred1)
print(cr)
```

```python
score1 = accuracy_score(ytest,y_pred1)
score1
```
