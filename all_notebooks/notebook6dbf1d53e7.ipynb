```python
#correlation map
f,ax = plt.subplots(figsize=(15, 15))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
#korekasyon haritasında, variable'ların birbiriyle olan alakasını görüyoruz. En alakalı olan variable'lar için
#'ladder score' ve 'upperwhisker' gösterilebilir. Korelasyonu en düşük olan variable'lar ise 'perceptions of corruption'
#ve 'explained by: perceptions of corruption' gösterilebilir. (neden olduğunu ben de bilmiyorum.)
```

```python
#Line plot for upperwhisker & lowerwhisker columns.
data.upperwhisker.plot(kind='line', color = 'b',label = 'upperwhisker',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
data.lowerwhisker.plot(kind='line', color = 'r',label = 'lowerwhisker',linewidth=1,alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')     
plt.xlabel('x axis')             
plt.ylabel('y axis')
plt.title('Line Plot')
#line plot'ta değerler arttıkça bir düşüş görüyoruz her iki variable için de. Bu açıdan doğru orantılılar,
#hatta en son geldikleri nokta bile aynı. sadece 40'a yaklaşırken ters orantılı bir hareket içerisinde
#olmuşlar.
```

```python
#scatter plot
data.plot(kind='scatter', x='upperwhisker', y='lowerwhisker',alpha = 0.5,color = 'orange')
plt.xlabel('upperwhisker')
plt.ylabel('lowerwhisker')
plt.title('upperwhisker & lowewhisker Scatter Plot')
#upperwhisker oranı belirli bir orana kadar arttıkça lowerwhisker oranı da artış göstermiş, fakat bu orandan
#uzaklaşıldıkça bu korelasyon yoğunluğu azalmaya başlamış.
```
