```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
covid=pd.read_csv('/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv')
```

```python
covid.head()
```

```python
covid.info()
```

```python
country=list(covid.country.unique())
print(country)
```

```python
covid[['country','vaccines']].drop_duplicates().set_index('country')
```

```python
covid.columns[[(0,2,3,4,5,8,9,10,11)]]
```

```python
covid_vaccinations=covid[covid.columns[0:12]].groupby('country').max()
covid_vaccinations.drop(['iso_code','date','daily_vaccinations_raw','daily_vaccinations'],axis=1,inplace=True)
covid_vaccinations.reset_index(inplace=True)
covid_vaccinations.fillna(0,inplace=True)
```

```python
covid_vaccinations.info()
```

```python
k=covid_vaccinations[covid_vaccinations.total_vaccinations>0].sort_values(by='people_vaccinated',ascending=False)
k.tail()
```

```python
plt.figure(figsize=(16,16))
k.sort_values('total_vaccinations',ascending=False,inplace=True)
sns.barplot(x=k.total_vaccinations,y=k.country,data=k)
plt.grid()
```

```python
plt.figure(figsize=(16,16))
k=covid_vaccinations[covid_vaccinations.total_vaccinations_per_hundred>0].sort_values(
    by='total_vaccinations_per_hundred',ascending=False)

sns.barplot(x=k.total_vaccinations_per_hundred,y=k.country,data=k)
plt.grid()
```

```python
plt.figure(figsize=(16,16))
k=covid_vaccinations[covid_vaccinations.people_fully_vaccinated>0].sort_values(by='people_vaccinated',ascending=False)

k.sort_values(by='people_fully_vaccinated',ascending=False,inplace=True,)
splot=sns.barplot(x=k.people_fully_vaccinated,y=k.country,data=k)

plt.grid()
```

```python
print('countries with no first stage of vaccination done ')

not_vaccinated=list(covid_vaccinations[covid_vaccinations.people_vaccinated==0]['country'])
not_vaccinated
```

```python
print('countries with  no first and secondstage vaccination done')

not_fully_vaccinated=list(covid_vaccinations[covid_vaccinations.people_fully_vaccinated==0]['country'])
not_fully_vaccinated
```

```python
print('countries with first stage of vaccination done for some part of population but no secondstage')
no_second_yesfirst=[i  for i in not_fully_vaccinated if i not in not_vaccinated]
no_second_yesfirst
```

```python
covid_vaccinations[covid_vaccinations.total_vaccinations_per_hundred>20]
```

```python

sns.distplot(covid_vaccinations.total_vaccinations_per_hundred)
plt.axvline(covid_vaccinations.total_vaccinations_per_hundred.mean(),color='red',label='mean')
plt.axvline(covid_vaccinations.total_vaccinations_per_hundred.median(),color='blue',label='median')
plt.legend()
```

```python
print('mean average vaccination per 100 person in 800 countries')
covid_vaccinations.total_vaccinations_per_hundred.mean()
```

```python
covid_vaccinations.total_vaccinations_per_hundred.median()
```

```python
from scipy.stats import zscore
sns.distplot(zscore(covid_vaccinations.total_vaccinations_per_hundred))
```

```python
print('countries 2.5 more away ')
covid_vaccinations[zscore(covid_vaccinations.total_vaccinations_per_hundred)>2.5][['country','total_vaccinations_per_hundred']]
```

```python
df=covid.copy()
```

```python
df['country_supplier']=df['country']+df['vaccines']
```

```python
df_country_count=pd.DataFrame(df['country_supplier'].value_counts())
df_country_count.reset_index(inplace=True)
df_country_count.rename(columns={'index':'country_supplier','country_supplier':'count'},inplace=True)
df_country_count
```

```python
plt.figure(figsize=(16,16))
sns.barplot(data=df_country_count,x='count',y='country_supplier')
plt.grid()
plt.show()
```
