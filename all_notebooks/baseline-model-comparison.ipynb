```python
df = pd.read_csv('../input/android-permission-dataset/Android_Permission.csv')
df.sample(5)
```

```python
# import a classification module of PyCaret
from pycaret.classification import *
```

```python
# select numeric columns as features
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

numeric_col = df.select_dtypes(include=numerics)
```

```python
numeric_col
```

```python
numeric_col.columns[numeric_col.isna().any()].tolist()
```

## Check Distribution of column with missing values

```python
numeric_col['Dangerous permissions count'].hist()
```

## Fill missing values with mean

```python
numeric_col['Dangerous permissions count'].fillna((numeric_col['Dangerous permissions count'].mean()), inplace=True)
```

## Setup Pycaret

```python
clf = setup(data = numeric_col, target = 'Class', silent = True, session_id = 123, train_size = 0.7)
```
