```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../input/zomato-bangalore-restaurants/zomato.csv')
```

```python
df.info()
```

```python
df.drop_duplicates(inplace = True)
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(), cbar=False, yticklabels = False)
```

```python
df.drop(['url','address','menu_item','reviews_list','phone','dish_liked','listed_in(city)'],inplace=True,axis =1)
```

```python
df.head()
```

```python
df = df.rename(columns={'approx_cost(for two people)':'cost','listed_in(type)':'type','listed_in(city)':'city'})
df.columns
```

```python
#Converting Cost dtypes
df['cost'] = df['cost'].astype(str)
df['cost'] = df['cost'].apply(lambda x: x.replace(',','')) 
df['cost'] = df['cost'].astype(float)
df['cost'].tail()
```

```python
df['rate']=df['rate'].replace('NEW','0')
df['rate']=df['rate'].replace('-','0')
df['rate']=df['rate'].str.replace('/5','').astype(float)
```

```python
df.dropna(inplace = True)
```
