## Machine Learning

Contrairement aux exercices précédents (diabète, fraude, zalando), nous ne devons plus prédire une classe (problèmes de classification) mais une valeur numérique, un prix. C'est un problème de régression.

Nous devons donc utiliser d'autres modèles que ceux vus précédemment, qui étaient destinés à la classification. Pour la plupart, nous allons simplement utiliser leur pendant adapté à la régression : le svc devient svr, la régression logistique devient régression linéaire, k-nn classifier devient k-nn reg,... même xgBoost et RandomForest ont maintenant leur équivalent en régression.

Qui dit changement de problème, dit également changement de méthode d'évaluation des modèles. La matrice de corrélation et l'accuracy ne sont plus utilisables dans un problème de régression. Nous utiliserons donc ici le score R2.

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

### Définition X/Y et Train/Test 

```python
X = df.drop(['MEDV'], axis = 1)
y = df.MEDV
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=1)
```

Régression Linéaire

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

SVR

```python
svr = SVR()
svr.fit(X_train, y_train)
svr.score(X_test, y_test)
```

Random Forest Regressor

```python
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

R2 est compris entre O et 1. Il tend vers 1 quand le modèle explique bien la distribution des points, et vers 0 lorsque modèle n'explique absoluement pas la distribution.
