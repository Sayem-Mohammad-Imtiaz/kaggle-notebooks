```python
import pandas as pd 

movies = pd.read_csv("/kaggle/input/imdb-extensive-dataset/IMDb movies.csv",low_memory=False)

#print(movies.shape)
```

```python
directros_actors_films = movies[['director', 'actors', 'title']]

data = {'film': [],
        'director': [],
        'actor': []}
df = pd.DataFrame(data)

for i in range(40, 50):#directros_actors_films size
    dirs_str = directros_actors_films.loc[i].loc['director']
    if pd.isna(dirs_str) == False:
        dirs_series = pd.Series(dirs_str.split(','))
        
        for j in range(dirs_series.size):
            actors_str = directros_actors_films.loc[i].loc['actors']
            if pd.isna(actors_str) == False:
                actors_series = pd.Series(actors_str.split(','))
                
                for k in range(actors_series.size):
                    new_row = {'film': directros_actors_films.loc[i].loc['title'], 
                               'director':dirs_series[j], 
                               'actor': actors_series[k]}
                    df = df.append(new_row, ignore_index=True)
#                     print(i)
#                     print('TITLE ' + directros_actors_films.loc[i].loc['title'] +
#                          ' DIRECTOR ' + dirs_series[j] + 
#                          ' ACTOR ' + actors_series[k])
    
    
df.groupby(['director', 'actor']).size().sort_values(ascending=False).head(20)
```
