# Import Labiraries

```python
import numpy as np
import pandas as pd
```

```python
#Loading the PlayTennis data
Play_Tennis = pd.read_csv('../input/playtennis/PlayTennis.csv')
```

```python
Play_Tennis
```

# Convert Classes to Numerical Values

```python
from sklearn.preprocessing import LabelEncoder
```

```python
#Create Object from LabelEncoder
label_En = LabelEncoder()
```

```python
Play_Tennis['outlook']= label_En.fit_transform(Play_Tennis['outlook'])
Play_Tennis['temp']=label_En.fit_transform(Play_Tennis['temp'])
Play_Tennis['humidity']=label_En.fit_transform(Play_Tennis['humidity'])
Play_Tennis['windy']=label_En.fit_transform(Play_Tennis['windy'])
Play_Tennis['play']=label_En.fit_transform(Play_Tennis['play'])
```

```python
Play_Tennis
```

# Standardization (Scaler)

```python
#Import Library StandardScaler
from sklearn.preprocessing import StandardScaler
#Creat Object from StandardScaler
s = StandardScaler()
#fit_transform for dataset
data_standardscaler = s.fit_transform(Play_Tennis)
```

```python
data_standardscaler
```

```python
#Import Library MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
#Creat Object from MinMaxScaler
s = MinMaxScaler()
#fit_transform for dataset
data_minmaxscaler = s.fit_transform(Play_Tennis)
```

```python
data_minmaxscaler
```

```python
#Import Library RobustScaler
from sklearn.preprocessing import RobustScaler
#Creat Object from RobustScaler
s = RobustScaler()
#fit_transform for dataset
data_robustscaler = s.fit_transform(Play_Tennis)
```

```python
data_robustscaler
```

```python
#Import Library Normalizer
from sklearn.preprocessing import Normalizer
#Creat Object from Normalizer
s =  Normalizer()
#fit_transform for dataset
data_normalizer = s.fit_transform(Play_Tennis)
```
