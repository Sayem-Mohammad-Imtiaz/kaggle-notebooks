```python
def get_index_from_name(name):
    return df[df["title"]==name].index.tolist()[0]
```

```python
def print_similar_animes(query=None):
     if query:
        found_id = get_index_from_name(query)
        for id in indices[found_id][1:]:
            print(df.iloc[id]["title"])
```
