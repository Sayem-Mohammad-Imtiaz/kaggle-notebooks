```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
team = pd.read_csv("../input/csgo-player-and-team-stats/team_stats.csv")
```

```python
team.drop("Unnamed: 0", axis = 1, inplace=True)
```

```python
team.head()
```

```python
highest_10_kd = team.sort_values("kd",ascending = False)
```

```python
highest_10_kd
```

```python
highest_10_kd.columns
```

```python
highest = highest_10_kd.iloc[:10]
```

```python
highest
```

```python

sns.pointplot(data=highest, x='name', y='kd', color='red');
plt.title("Team Having Higher no of KD")
plt.tick_params(axis='x', which='major', labelsize=15, rotation=90)
plt.tick_params(axis='y', which='major', labelsize=15)
```

# Most Experienced Team

```python
most_exp = team.sort_values("total_maps",ascending = False)
```

```python
most_exp
```

```python
most_exp_10 = most_exp.iloc[:10]
```

```python
most_exp_10
```

```python
sns.pointplot(data=most_exp_10, x='name', y='total_maps', color='red');
plt.title("MOST EXPERIENCED TEAM")
plt.tick_params(axis='x', which='major', labelsize=15, rotation=90)
plt.tick_params(axis='y', which='major', labelsize=15)
```

```python
plt.figure(figsize=(20,12))
sns.barplot(data=most_exp_10, y="name", x="total_maps")
plt.title('most maps played by Teams')
plt.xlabel('TOTAL MAPS PLAYED')
plt.ylabel('NAME OF TEAM')
plt.show()
```

```python
rating = team.sort_values("rating", ascending = False)
```

```python
rating 
```

```python
rating_10 = rating.iloc[:10]
```

```python
rating_10
```

```python
sns.pointplot(data=rating_10, x='name', y='rating', color='red');
plt.title("Top Ratings By Team")
plt.tick_params(axis='x', which='major', labelsize=15, rotation=90)
plt.tick_params(axis='y', which='major', labelsize=15)
```
