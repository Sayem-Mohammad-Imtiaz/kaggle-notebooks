# **IMPORTING LIBRARIES**

```python
from tensorflow.keras import models,layers
from tensorflow.keras.regularizers import l1

data=pd.read_csv('/kaggle/input/ionosphere01/ionosphere_data.csv')
```

```python
data
```

# **CHECKING NULL VALUES**

```python
data.isnull().sum()
```

# **DROPPING DUPLICATES,IF EXIST**

```python
data.drop_duplicates(inplace=True)
```

# **SHUFFLING DATA ROWS**

```python
data=data.sample(frac=1)
```

# **ENCODING label COLUMN & NORMALISING DATA**

```python
data['label']=data['label'].map({'g':1,'b':0})
labels=data['label']

data.drop('label',inplace=True,axis=1)

mean=data.mean(axis=0)
std=data.std(axis=0)
data-=mean
data/=mean
```

```python
data.head()
```

# ASSIGNING VALUE 0 TO feature2 AS IT WAS DEFAULT

```python
data['feature2']=0
```

# SPLITTING DATA INTO TRAIN,TEST

```python
train_data=data[:210]
train_labels=labels[:210]

test_data=data[210:]
test_labels=labels[210:]
```

# CREATING MODEL

```python
model=models.Sequential()
model.add(layers.Dense(16,activation='relu',input_shape=(train_data.iloc[1].shape),kernel_regularizer=l1(0.001)))
model.add(layers.Dropout(0.3))
model.add(layers.Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
data_history=model.fit(train_data,train_labels,epochs=100,batch_size=2)
```

# EVALUATING ON TEST DATA

```python
model.evaluate(test_data,test_labels)
```

# PREDICTING ONE ROW

```python
predictions=model.predict(test_data)
predictions[88]
```
