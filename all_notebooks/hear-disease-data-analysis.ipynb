```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import SelectKBest, f_classif

dataset = pd.read_csv('../input/heart-disease-cleveland-uci/heart_cleveland_upload.csv')

plt.figure(figsize=(12,10))
sns.heatmap(dataset.corr(),annot=True,cmap="magma",fmt='.2f')

drop_cols=['condition']
X = dataset.drop(drop_cols, axis = 1) # X = independent variables
y = dataset['condition'] # y = target column 

best_features = SelectKBest(score_func=f_classif, k=10)
fit = best_features.fit(X,y)
df_scores = pd.DataFrame(fit.scores_)
df_columns = pd.DataFrame(X.columns)

feature_scores = pd.concat([df_columns, df_scores],axis=1)
feature_scores.columns = ['Feature_Name','Score'] 
print(feature_scores.nlargest(10,'Score'))  
```
