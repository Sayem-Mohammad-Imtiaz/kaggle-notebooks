```python
import pandas as pd
```

```python
import plotly.express as px
```

```python
import numpy as np
```

```python
import plotly.figure_factory as ff
```

```python
df=pd.read_csv('../input/starbucks-customer-retention-malaysia-survey/Starbucks satisfactory survey.csv')
```

```python
df.rename(columns={"1. Your Gender": 'Gender',
                                "2. Your Age": "ageGroup",
                                "3. Are you currently....?":"status",
                                "4. What is your annual income?":"annualIncome",
                                "5. How often do you visit Starbucks?":"visitFrequency",
                                "6. How do you usually enjoy Starbucks?":"Method",
                                "7. How much time do you normally  spend during your visit?":"timeCost",
                                "8. The nearest Starbucks's outlet to you is...?":"location",
                                "9. Do you have Starbucks membership card?":"membership",
                                "10. What do you most frequently purchase at Starbucks?":"mostPurchase",
                                "11. On average, how much would you spend at Starbucks per visit?":"moneySpend",},inplace=True)
```

```python
df.columns
```

```python
tempdf=df.groupby('ageGroup').size().reset_index(name='count').sort_values('count',ascending=False)
```

# Visits by age

```python
fig=px.bar(tempdf,x='ageGroup',y='count',title="Visits by age",labels={'ageGroup':"Age Group",'count':"Number of visits"},text='count',height=500,width=600)
fig.update_traces(textposition='outside')
fig.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')
fig.show()
```

# Which foods contibute to more revenue.?

```python
df['moneySpend']=df['moneySpend'].replace({'Zero':0,'Less than RM20':10,'Around RM20 - RM40':30,'More than RM40':50})
```

```python
pur_mon_df=df.groupby(['mostPurchase']).sum()['moneySpend'].reset_index(name='total').sort_values('total',ascending=False).head(5)

# .size().rename(columns={'size':"count"}).sort_values('count',ascending=False).head(5)
```

```python
fig=px.bar(pur_mon_df,y='mostPurchase',x='total',title="Top 5 revenue foods",text='total')
fig.update_traces(textposition='outside')
fig.update_xaxes(tickprefix="RM")
fig.show()
```

# Does having a membership card affects revenue.?

```python
mem_mon_df=df.groupby(['membership']).sum()['moneySpend'].reset_index(name='total')
```

```python
fig=px.bar(mem_mon_df,y='total',x='membership',title="Revenue with & without membership",text='total',color='membership',height=500,width=500)
fig.update_traces(textposition='outside')
fig.update_yaxes(tickprefix="RM")
fig.show()
```

# Mean time of groups visting the starbucks

```python
df['timeCost']=df['timeCost'].replace({'Between 2 hours to 3 hours':np.round(np.random.uniform(2,3),2),'Below 30 minutes':np.round(np.random.uniform(0,0.5),2),'More than 3 hours':np.round(np.random.uniform(3,6),2),'Between 30 minutes to 1 hour':np.round(np.random.uniform(0.5,2)),'Between 1 hour to 2 hours':np.round(np.random.uniform(1,2),2)}).reset_index()
```

```python
mean_group_df=df.groupby('status')['timeCost'].mean().reset_index(name='total time')
mean_group_df['total time']=mean_group_df['total time'].round()
```

```python
fig=px.bar(mean_group_df,y='total time',x='status',title="Average time spent by different groups",text='total time',color='status',height=600,width=600)
fig.update_traces(textposition='outside')
fig.update_yaxes(ticksuffix=" Hours")
fig.show()
```
