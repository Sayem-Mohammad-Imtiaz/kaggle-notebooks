```python
from wordcloud import WordCloud, STOPWORDS
```

```python
df = pd.read_csv('/kaggle/input/tamil-movies-dataset/Tamil_movies_dataset.csv')
df
```

```python
df['club'] = df['Actor'] + ' '+df['Director'] + ' '+df['Genre']
df['club']
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(ngram_range=(1, 2),stop_words='english',min_df=0)
count_matrix = count.fit_transform(df['club'])
count_matrix
```

```python
indices = pd.Series(df.index, index=df['MovieName'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def get_recommendations(title):
    
    idx = indices[title]

    co_similar = cosine_similarity(count_matrix, count_matrix)

    similar_scores = list(enumerate(co_similar[idx]))

    similar_scores = sorted(similar_scores, key=lambda x: x[1], reverse=True)
    
    xx = 8
    xx+=1

    similar_scores = similar_scores[1:xx]

    movie_indices = [i[0] for i in similar_scores]

   
    mv =  df['MovieName'].iloc[movie_indices]
    
    cld = str(mv)
    
    # mv.reset_index(drop=True, inplace=True)
    
    for i in cld:
        r_cld = cld.replace(" ", "")
    
    wc = WordCloud(background_color='black',max_font_size=40,contour_width=3)
    
    wc.generate(r_cld)
    plt.imshow(wc, interpolation='bilinear')
    plt.axis("off")
    plt.show()
    
    return mv
```

```python
m_name = str('Bigil')
get_recommendations(m_name)
```
