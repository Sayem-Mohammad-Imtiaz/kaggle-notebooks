Understanding few Amenities
1. **Annapurna Meals: Addressing the Urban Hunger**

GHMC provides hygienic and safe Food to under privileged, poor and down trodden, at just Rs.5/- per meal. The scheme had started with 8 centres in 2014 and has now expanded to 150 centre’s serving 25,000 people with hot and nourishing meals daily through Hare Krishna Movement Charitable Foundation. There are plans to extend the number of centres to meet the burgeoning demand for these centres owing to its huge popularity among the poor. In this scheme Rs. 5/- is paid by the beneficiary and GHMC subsidizes remaining amount to Hare Krishna Movement Charitable Foundation. (Info from https://csr.ghmc.gov.in/feedingthepoor.aspx )
![Annapurna Canteen](https://csr.ghmc.gov.in/images/InnerImages/Annapurna.jpg)


2. **NIGHT SHELTERS**

The aim is to see that no one has to fight the cold out on the streets. They can go to our shelters and stay there. We will have mattresses, blankets and drinking water in all our shelters. Currently, there are 15 night shelters in the city and GHMC operates these facilities in association with different NGOs and voluntary organisations. This apart, the municipal corporation distributes blankets for the poor and homeless in many areas to help them stay warm.Ref(https://telanganatoday.com/ghmc-to-set-up-more-night-shelters-to-keep-homeless-warm)

3. **ANGANWADI CENTERS**

Anganwadi is a type of rural child care centre in India. They were started by the Indian government in 1975 as part of the Integrated Child Development Services program to combat child hunger and malnutrition. Anganwadi means "courtyard shelter" in Hindi.


4.**2BHK**

The Telangana government introduced the Dignity Housing Scheme, popularly called the 2BHK Housing Scheme or the double room scheme in October 2015, to ensure that those in need of a roof over their head while being unable to afford it, can be eligible for a property under this scheme. While the centre and various state governments also offer subsidised housing for the economically weaker sections, the Telangana government went a step further, to make these units, priced anywhere between Rs 5 lakhs to Rs 8.65 lakhs, totally free of cost for the beneficiaries. The housing department has taken up the construction of 2.80 lakh units.
![2BHK](http://haigreeva.com/wp-content/uploads/2018/06/Ameenpur-1.jpeg) (Ref: haigreeva.com)

```python

# Create a map
m_1 = folium.Map(location=[17.400789,78.400245], tiles='cartodbpositron', zoom_start=11.25)

# Display the map
# Add points to the map
count = 0
for idx, row in GHMC.iterrows():
    Marker([row['Latitude'], row['Longitude']],icon=folium.Icon(color=ID_Mappings[row['Category']][0], 
                                                                icon=ID_Mappings[row['Category']][1], 
                                                                prefix='fa',
                                                               ),
          tooltip=row['Category']
          ).add_to(m_1)
    count+=1
    if count == 10:
        break
choropleth = Choropleth(geo_data=ghmc ,columns=['Ward_Name']
          ).add_to(m_1)
choropleth.geojson.add_child(
    folium.features.GeoJsonTooltip(['Ward_Name'],labels=False)
)
m_1
```

```python
finding_ward_name = []
real_ward_name, real_ward_number = [],[]
for point in ghmc_amenities.geometry.values:
    count = 0
    for idx, row in ghmc.iterrows():
        if point.within(row.geometry):
            finding_ward_name.append(row.geometry)
            real_ward_number.append(row.Ward_Number)
            real_ward_name.append(row.Ward_Name)
            count = 0
            break
        count+=1
    if count == len(ghmc.geometry.values):
        finding_ward_name.append(np.nan)
```

```python
ghmc_amenities["WARD_NAME"] = finding_ward_name
ghmc_amenities.head()
```

```python
ghmc_amenities.dropna(subset = ['WARD_NAME'], inplace=True)
ghmc_amenities.head()
```

```python
ghmc_amenities["Ward_Number"] = real_ward_number
ghmc_amenities["Ward_Name"] = real_ward_name
ghmc_amenities.head()
```

```python
ghmc.sort_values("Ward_Number").head()
```

```python
Ward_Wise_Amenities = ghmc_amenities.groupby(["Ward_Name","Category"])["Sno"].count().reset_index()
Ward_Wise_Amenities.columns = ["Ward_Name","Amenties","Count"]
def fetch_ward_number(value):
    ward_name_to_number = dict(zip(ghmc["Ward_Name"],ghmc["Ward_Number"]))
    return ward_name_to_number[value]
Ward_Wise_Amenities["Ward_Number"] = Ward_Wise_Amenities.Ward_Name.apply(fetch_ward_number)

Ward_Wise_Amenities.head()
```

```python
Wards_having_hospitals = Ward_Wise_Amenities[Ward_Wise_Amenities.Amenties == 'HOSPITALS']
Wards_having_hospitals.sort_values("Count").head()
```

```python
print("Wards that doesn't have even one hospital")
ward_name_to_number.keys() - Wards_having_hospitals.Ward_Name.values
```

```python
list_of_Amenities = set(Ward_Wise_Amenities.Amenties.values.tolist())
list_of_wards = set(ward_name_to_number.keys())
```

```python
Missing_Amenities = {}
for Ward in list_of_wards:
    Missing_Amenities[Ward] = list_of_Amenities - set(Ward_Wise_Amenities[Ward_Wise_Amenities.Ward_Name == Ward].Amenties.values.tolist())
```

```python
len(Missing_Amenities)
```

```python
len(list_of_wards)
```

```python
Wards_without_Amenities = pd.DataFrame(columns=["Ward_Name","Amenities"])
for key,value in Missing_Amenities.items():
    for item in value:
        Wards_without_Amenities = Wards_without_Amenities.append({"Ward_Name" : key, "Amenities":item}, ignore_index = True)
no_of_Amenities_missed = Wards_without_Amenities.groupby(["Ward_Name"]).count().reset_index()

print("List of no.of Amenties that aren't in respective Ward")
no_of_Amenities_missed.sort_values("Amenities")
```
