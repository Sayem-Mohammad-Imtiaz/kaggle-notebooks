# Importing Relevant Libraries

```python
#Data Analysis
import pandas as pd
import numpy as np
#Data Visualisation
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
#Data Imputing
data=pd.read_csv('../input/imdb-5000-movie-dataset/movie_metadata.csv')
data.head(5)
```

```python
print('Number of rows:',data.shape[0])
print('Number of columns:',data.shape[1])
```

```python
data.info()
```

```python
#check for null values
(data.isnull().sum()/len(data)*100)
```

```python
data.duplicated().any()
```

```python
data=data.drop_duplicates()
data.shape
```

#Top 3 columns with most missing values : 'gross','budget, 'aspect_ratio'

#1.Year with highest release of movies

```python
data['title_year'].value_counts().reset_index().sort_values(by='title_year',ascending=False).head(5)
```

```python
plt.figure(figsize=(20,5))
sns.set(style='dark')
sns.countplot(data['title_year'].sort_values(ascending=False))
plt.xticks(rotation=90)
plt.show()
```

#Highest number of movies were released in the year 2009 followed by 2014 and 2006.

#2. Movies with and IMDb rating of 9 and above

```python
len(data[data['imdb_score']>=9])
```

#Less than 10 movies were rated 9.0 or above 

#3. Distribution of movies wrt ratings

```python
plt.figure(figsize=(10,5))
sns.set(style='dark')
sns.histplot(data=data,x='imdb_score',color='indigo',binwidth=1)
```

#4. Movies with highest and lowest budget

```python
print("Movie with minimum budget:")
min_bud=data[data['budget']==data['budget'].min()][['movie_title','budget']]
min_bud
```

```python
print("Movie with maximum budget:")
max_bud=data[data['budget']==data['budget'].max()][['movie_title','budget']]
max_bud
```

#5. Movies with smallest and longest runtime

```python
data[data['duration']==data['duration'].min()][['movie_title','duration','genres','language','imdb_score']]
```

```python
data[data['duration']==data['duration'].max()][['movie_title','duration','genres','language','imdb_score']]
```

#6. Most liked genres

```python
#filtering out movies with rating 8.5 and above
most_liked=data[data['imdb_score']>=8.5]
most_liked_per=(most_liked['genres'].value_counts())/(data['genres'].value_counts())*100
most_liked_per.sort_values(ascending =False).head(10)
```

#7. Movie with Highest And Lowest Votes

```python
#defining a function to get highest and lowest values
def max_and_min(x):
    min_index=data[x].idxmin
    max_index=data[x].idxmax
    print("Movie which Has Highest "+ x + " : ",data['movie_title'][max_index])
    print("Movie which Has Lowest "+ x + " : ",data['movie_title'][min_index])
```

```python
max_and_min('num_voted_users')
```

```python
#8. Which length movies most liked by the audiences according to their imdb_score?
liked_movies=pd.DataFrame(data.groupby(['duration']).mean()['imdb_score'])
```

```python
plt.figure(figsize=(16,5))
sns.set(style='darkgrid')
sns.lineplot(data=liked_movies,x='duration',y='imdb_score',color='orange')
```

#9.Average Runtime Of Movies From Year To Year

```python
mean_duration=pd.DataFrame(data.groupby('title_year').mean()['duration'])
```

```python
plt.figure(figsize=(16,5))
sns.set(style='whitegrid')
sns.lineplot(data=mean_duration,x='title_year',y='duration',color='maroon')
```
