```python
def outlier_plot(data,feat_1):
    plt.figure(figsize=[13,13])
    plt.subplot(2,2,1)
    sbn.boxplot(feat_1,data=data,palette='Set3')
    
    plt.subplot(2,2,2)
    data[feat_1].plot.hist(density=True,edgecolor='black',color=sbn.color_palette('cubehelix_r'))
    plt.suptitle('Distribution of '+feat_1,fontsize=25)
    plt.show()
```

```python
outlier_plot(data,'ssc_percentage')
outlier_plot(data,'hsc_percentage')
outlier_plot(data,'degree_percentage')
outlier_plot(data,'etest_percentage')
outlier_plot(data,'mba_percentage')
```

# Feature Engineering

```python
data.head(2)
```

### As machine understands the data in numerical format we need to convert categorical data into numerical data.So,let's start with it

```python
data['gender'] = np.where(data['gender']=='M',1,0)
data['ssc_board'] = np.where(data['ssc_board']=='Others',1,0)
data['hsc_board'] = np.where(data['hsc_board']=='Others',1,0)
data['workex'] = np.where(data['workex']=='Yes',1,0)
data['specialisation'] = np.where(data['specialisation']=='Mkt&HR',1,0)
data['status'] = np.where(data['status']=='Placed',1,0)
```

```python
data.head()
```

```python
for column in ['hsc_stream', 'degree_type']:
    dummies = pd.get_dummies(data[column],drop_first=True)
    data[dummies.columns] = dummies
```

```python
data.drop(['degree_type','hsc_stream'], axis=1, inplace=True)
```

```python
data.head()
```

```python

plt.figure(figsize=(10,5))
sbn.distplot(data['salary'], bins=50, hist=False,color='#3e6870')
plt.title("Salary Distribution")
plt.show()
```

```python
data1 = data.drop('salary',axis=1)
data1.head()
```

```python
x = data1.drop('status',axis=1)
y = data1['status']
```

```python
x.head()
```

# Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x = pd.DataFrame(scaler.fit_transform(x),columns=x.columns)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=101)
```

# Applying ML algorithms

## Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression(random_state=0)
logreg.fit(x_train,y_train)
```

```python
y_pr = logreg.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pr))
print(metrics.confusion_matrix(y_test,y_pr))
print(metrics.classification_report(y_test,y_pr))
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state=0)
rfc.fit(x_train,y_train)
```

```python
y_pre = rfc.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pre))
print(metrics.confusion_matrix(y_test,y_pre))
print(metrics.classification_report(y_test,y_pre))
```

## K-Neighbors Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knc = KNeighborsClassifier()
knc.fit(x_train,y_train)
```

```python
y_pred = knc.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,y_pred))
print(metrics.confusion_matrix(y_test,y_pred))
print(metrics.classification_report(y_test,y_pred))
```

```python
error_rate=[]
for i in range(1,150,5):
    knc = KNeighborsClassifier(n_neighbors=i)
    knc.fit(x_train,y_train)
    pred_i = knc.predict(x_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,150,5),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knc = KNeighborsClassifier(n_neighbors = 20)
knc.fit(x_train,y_train)
y_predd = knc.predict(x_test)
from sklearn import metrics
print(metrics.confusion_matrix(y_test,y_predd))
print(metrics.accuracy_score(y_test,y_predd))
print(metrics.classification_report(y_test,y_predd))
```

## Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier() 
dtc.fit(x_train,y_train)
```

```python
y_predi = dtc.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test,y_predi))
print(metrics.accuracy_score(y_test,y_predi))
print(metrics.classification_report(y_test,y_predi))
```

```python
from sklearn import tree
plt.figure(figsize=[15,15])
tree.plot_tree(dtc,filled=True)
plt.show()
```

```python
path = dtc.cost_complexity_pruning_path(x_train,y_train)
ccp_alphas , impurities = path.ccp_alphas , path.impurities
```

```python
print("ccp_alphas :- \n",ccp_alphas)
print("impurities :- \n",impurities)
```

```python
clfs = []
for ccp_alpha in ccp_alphas:
    dct = DecisionTreeClassifier(random_state=0,ccp_alpha = ccp_alpha)
    dct.fit(x_train,y_train)
    clfs.append(dct)
print("Number of nodes in the last tree is: {} with ccp_alpha: {}".format(
      clfs[-1].tree_.node_count, ccp_alphas[-1]))
```

```python
train_scores = [dct.score(x_train,y_train) for dct in clfs]
test_scores = [dct.score(x_test,y_test) for dct in clfs]
```

```python
fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker='o', label="train",
        drawstyle="steps-post",color='red')
ax.plot(ccp_alphas, test_scores, marker='o', label="test",
        drawstyle="steps-post")
ax.legend()
plt.show()
```

```python
dct = DecisionTreeClassifier(random_state=0,ccp_alpha=0.018)
dct.fit(x_train,y_train)
```

```python
y_predict = dct.predict(x_test)
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test,y_predict))
print(metrics.accuracy_score(y_test,y_predict))
print(metrics.classification_report(y_test,y_predict))
```
