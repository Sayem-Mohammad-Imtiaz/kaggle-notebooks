using panda we can read the csv and using head() we can print out the first 5 rows of the csv

```python
wbc = pd.read_csv('/kaggle/input/wisconsin-breast-cancer-cytology-features/wisconsin_breast_cancer.csv')
wbc.head()
```

more info on the data

```python
wbc.info()
```

since 'nuclei' has some null value it better remove all the null data values from the original and create a new dataset 

because if not it gives us some error which we might want to avoid.

```python
new_data = wbc.dropna()

new_data.info()
```

plotting the data 

using the diag_kind as a "hist" since the hue has only 2 values.

```python
import seaborn as sns
sns.pairplot(data=new_data, hue="class", palette="Set2" ,diag_kind="hist")
```

In the above graphs orange indicated data corresponding to Class 1 and green with Class 0

Splitting th data into X and y ... X contains all features from column 1 to 10 and y contains "id" and "class".

```python
X = new_data.iloc[0:683 , 1:10]
print(X)
```

```python
y = new_data.iloc[0:683 , 10:11]
print(y)
```

splitting the X and y into train and test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=1502775)
print(X_test)
print(y_train)
```

the above data set represents the set which our model will learn so as to predict the class of out test data sets.  

Now that we have a X_train and y_train we make the program learn the data based on these 2 sets using the support vector classification 

```python
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)
```

Now that we have a model ready to we can test the set called X_test to predict its restective class in y_test.

```python
pred = model.predict(X_test)
print(pred[0:10])
print(y_test[0:10])
```

in the above result we can see that the model was able to predict the first 10 data values but to understand it better we can produce a confusion matrix and a classification report.

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test,pred))
```

in the above confusion matrix we can see that the diagnol matrix makes up for most of the data with is a good thing.

Generating the classification report of the test

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,pred))
```

The above test report states that the program had a 95% accuracy in finding out the class of each row of data that corresponds to their respective class.
