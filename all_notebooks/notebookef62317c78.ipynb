```python
import pandas as pd
import numpy as np
import seaborn as sns
import cufflinks as cf
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.graph_objs as go 
```

```python
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
init_notebook_mode(connected=True)
cf.go_offline()
from plotly import __version__
```

```python
df = pd.read_csv('../input/imf-weo-africa-20192020/IMF WEO Africa 2019-2020.csv',encoding = "ISO-8859-1")
```

```python
df.head(2)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Dropping all the 'Units' has have 'National currency' as we'll like our analyis to be done in US dollars
indexNames = df[df['Units']=='National currency'].index
```

```python
df.drop(indexNames,inplace=True)
```

```python
df['Units'].value_counts()
```

```python
df['Subject Descriptor'].value_counts()
```

```python
df.isnull().sum()
```

```python
# Converting the Data 2020 to a numeric column alowing us to sum it up.
df['Data 2020_1'] = df['Data 2020'].apply(pd.to_numeric, downcast='float', errors='coerce')
```

```python
df.info()
```

```python
df['Cur Acct Bal'] = df[df['Subject Descriptor']=='Current account balance']['Data 2020_1']
df['Pop'] = df[df['Units']=='Persons']['Data 2020_1']
df['GDP Curr Prices'] = df[df['Subject Descriptor']=='Gross domestic product, current prices']['Data 2020_1']

```

```python
df['Cur Acct Bal'].sum()
```

```python
# The 3 largest Population
df.nlargest(3,'Pop')
```

```python
# The 3 smallest Population
df.nsmallest(3,'Pop')
```

```python
df['Cur Acct Bal'].fillna( value=0,axis=0,inplace=True)
```

```python
df['GDP Curr Prices'].fillna( value=0,axis=0,inplace=True)
```

```python
df['Pop'].iplot(kind='hist')
```

```python
data = dict(type='choropleth',
            colorscale = 'Viridis',
            reversescale = True,
            locations = df['Country'],
            z= df['Pop'],
            locationmode = 'country names',
            text = df['Country'],
            colorbar ={'title':'Country Population'})
```

```python
layout = dict(title = 'IMF, WE Outlook Database 2020- Africa Population',
             geo = dict(showframe=False,
                        showcoastlines=False,
                       projection = {'type':'mercator'}))
```

```python
choromap3 = go.Figure(data=[data],layout=layout)
iplot(choromap3)
```

```python
data = dict(type='choropleth',
            colorscale = 'YlOrBr',
            reversescale = True,
            locations = df['Country'],
            z= df['Cur Acct Bal'],
            locationmode = 'country names',
            text = df['Country'],
            colorbar ={'title':'Cur Acct Bal'})
```

```python
layout = dict(title = 'IMF, WE Outlook Database 2020- Africa Cur Acct Bal ',
             geo = dict(showframe=False,
                        showcoastlines=False,
                       projection = {'type':'mercator'}))
```

```python
choromap3 = go.Figure(data=[data],layout=layout)
```

```python
# The IMF Current Balance for these countries is the meassure of their earnings and spending for the Year 2020.
iplot(choromap3)
```
