```python
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
import missingno as msno
testvr = pd.read_csv("../input/food-prices-in-turkey/test.csv")
trainvr = pd.read_csv("../input/food-prices-in-turkey/train.csv")
print(testvr.shape)
print(trainvr.shape)
```

```python
trainvr = pd.concat([trainvr,testvr])
```

```python
msno.matrix(trainvr);
```

```python
print(trainvr["Place"].unique())
print(trainvr["ProductName"].unique().shape)
print(trainvr["UmName"].unique())
```

```python
trainvr.describe()
```

```python
sbn.pairplot(trainvr);
sbn.despine(bottom=True,left=True);
```

```python
plt.subplots(figsize=(20,10))
sbn.set_theme(style=("whitegrid"))
sbn.countplot(trainvr.Place);
sbn.despine(bottom=True,left=True);
```

```python
sbn.set_theme(style=("whitegrid"))
plt.subplots(figsize=(20,10))
sbn.lineplot(hue="Year",y="Price",data=trainvr,x="Month");
sbn.despine(bottom=True,left=True);
```

```python
plt.subplots(figsize=(20,10))
sbn.set_theme(style=("whitegrid"))
sbn.lineplot(x="Year",y="Price",data=trainvr,hue="Place");
sbn.despine(bottom=True,left=True);
plt.legend(bbox_to_anchor=(1.3, 0., 0.1, 0.5));
```

```python
plt.subplots(figsize=(20,10))
sbn.set_theme(style=("whitegrid"))
sbn.lineplot(hue="Year",y="Price",data=trainvr,x="Place");
sbn.despine(bottom=True,left=True);
plt.legend(bbox_to_anchor=(1.1, 0., 0.1, 0.5));
```

```python
plt.subplots(figsize=(20,10))
sbn.set_theme(style=("whitegrid"))
sbn.scatterplot(hue="Year",y="Price",data=trainvr,x="ProductName");
sbn.despine(bottom=True,left=True);
plt.legend(bbox_to_anchor=(1.1, 0., 0.1, 0.5));
plt.xticks(rotation=90);
```

```python
plt.subplots(figsize=(20,10))
sbn.set_theme(style=("whitegrid"))
sbn.scatterplot(y="Place",data=trainvr,x="ProductName");
sbn.despine(bottom=True,left=True);
plt.legend(bbox_to_anchor=(1.1, 0., 0.1, 0.5));
plt.xticks(rotation=90);
```
