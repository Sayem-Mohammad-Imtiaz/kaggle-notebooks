Now, let's differentiate our Independent and Dependend Columns follwed by spliting them to test and train

```python
X = train_data.iloc[:,1:]
y = train_data.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 25)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression
regresor= LinearRegression()
regresor.fit(X_train, y_train)
regresor_pred = regresor.predict(X_test)
```

## Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor()
rf_reg.fit(X_train, y_train)
randomforest_pred= rf_reg.predict(X_test)
```

## Evaluation (Our Finale)

### Explained variance regression score 

Best possible score is 1.0, lower values are worse.

```python
from sklearn.metrics import explained_variance_score
print(explained_variance_score(y_test, regresor_pred))
print(explained_variance_score(y_test, randomforest_pred))
```

### Mean absolute error regression loss
[https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_absolute_error.html](http://)

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(y_test, regresor_pred))
print(mean_absolute_error(y_test, randomforest_pred))
```

## Mean squared error regression loss
[https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html#sklearn.metrics.mean_squared_error](http://)

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, regresor_pred))
print(mean_squared_error(y_test, randomforest_pred))
```

## Median absolute error regression loss

Median absolute error output is non-negative floating point. The best value is 0.0

```python
from sklearn.metrics import median_absolute_error
print(median_absolute_error(y_test, regresor_pred))
print(median_absolute_error(y_test, randomforest_pred))
```

## R^2 (coefficient of determination) regression score

Best possible score is 1.0 and it can be negative (because the model can be arbitrarily worse). A constant model that always predicts the expected value of y, disregarding the input features, would get a R^2 score of 0.0.

[https://scikit-learn.org/stable/modules/generated/sklearn.metrics.r2_score.html#sklearn.metrics.r2_score](http://)

```python
from sklearn.metrics import r2_score
print(r2_score(y_test, regresor_pred))
print(r2_score(y_test, randomforest_pred))
```
