# creating choropleth map with life expectations for the year 2016

```python
map=folium.Map(location=[-15.788497,-47.879873],zoom_start=2,)
map
```

```python
#Setting up the world countries data URL
url = 'https://raw.githubusercontent.com/python-visualization/folium/master/examples/data'
country_shapes = f'{url}/world-countries.json'
```

```python
data2016 = dataDF[(dataDF['Year']==2016)]
data2016mod =data2016.replace('United States', "United States of America", inplace = True)
```

```python
#Adding the Choropleth layer onto our base map
folium.Choropleth(
    #The GeoJSON data to represent the world country
    geo_data=country_shapes,
    name='choropleth Life expectancy',
    data=data2016,
    #The column aceppting list with 2 value; The country name and  the numerical value
    columns=['Entity', 'Life expectancy'],
    key_on='feature.properties.name',
    fill_color='PuRd',
    nan_fill_color='white'
).add_to(map)
map
```
