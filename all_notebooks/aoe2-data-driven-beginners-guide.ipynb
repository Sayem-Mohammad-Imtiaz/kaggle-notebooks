# Win Rate of Civs in 1v1 ranked games in Age of Empire 2

This notebook is my first notebook written in Python and it is about the analysis of win rate of 37 civs in the game upto the Lord of the West DLC. This serves as a reference for rank players and also for game developers who aim to balance the game in later patch. 

I have retrieved the data in Kaggle and the raw data looks like:

```python
win_rate_table = pd.read_csv('/kaggle/input/35000-age-of-empires-2-1v1-ranked-random-matches/rankedrm.csv')
win_rate_table.head()
```

Each game takes place in a map, and the data records games from 12 different types of maps, where the map Arabia is the most popular among players.

```python

size = len(win_rate_table.index)
map_types = win_rate_table['map_type.name'].value_counts().reset_index().rename(columns={'index':'maps','map_type.name':'num_games'})
map_types.index += 1
map_types['percentage'] = map_types.num_games * 100 / size 

print(map_types)

map_types_plot = win_rate_table['map_type.name'].value_counts().plot.pie()
```

We can then analyse the win rate of each civ.

```python

        
#win_rate_table = pd.read_csv('/kaggle/input/35000-age-of-empires-2-1v1-ranked-random-matches/rankedrm.csv',usecols=['rating.win','rating.lose','map_type.name','civ.win.name','civ.lose.name']) 




        
win_rate_all = win_rate_table['civ.win.name'].value_counts()/(win_rate_table['civ.win.name'].value_counts()+win_rate_table['civ.lose.name'].value_counts())
win_rate_all = win_rate_all.sort_values(ascending = False).reset_index().rename_axis("rank", axis='columns').rename(columns={'index':'civs',0:'win_rate'})

num_civ = len(win_rate_all.index)
win_rate_all.win_rate = win_rate_all.win_rate * 100
win_rate_all.index += 1
win_rate_all['advantage'] = win_rate_all.win_rate - 50
        
print("In", size, "recorded AOE2 ranked 1v1 games, the win rates of the" ,num_civ, "civs are as follows: \n")


print(win_rate_all)
```

By observations, it is known that the behaviours of players in different rank varies a lot. It is also worth to analysis the win rate in different ranks. Here we use the convention low elo as ranking <1000, middle elo as 1000-1300 and high elo as > 1300. The following table summarizes the distribution:


```python
#Filter by elo
win_rate_table_low = win_rate_table.loc[win_rate_table['rating.win' ] < 1000]
size_low = len(win_rate_table_low.index)
win_rate_table_mid = win_rate_table.loc[win_rate_table['rating.win'] >= 1000]
win_rate_table_mid = win_rate_table_mid.loc[win_rate_table['rating.win'] < 1300]
size_mid = len(win_rate_table_mid.index)
win_rate_table_high = win_rate_table.loc[win_rate_table['rating.win'] >= 1300]
size_high = len(win_rate_table_high.index)


elos = pd.DataFrame({ 
              'num_games': [size_low, size_mid, size_high, size]},
             index=['low_elo', 'mid_elo','high_elo','total'])

print(elos)
elos_plot = elos.iloc[0:3].plot.pie(y = 'num_games')
```

```python
#analysis for low elo games


win_rate_low = win_rate_table_low['civ.win.name'].value_counts()/(win_rate_table_low['civ.win.name'].value_counts()+win_rate_table_low['civ.lose.name'].value_counts())
win_rate_low = win_rate_low.sort_values(ascending = False).reset_index().rename_axis("rank", axis='columns').rename(columns={'index':'civs',0:'win_rate'})
'''
win_rate_low = win_rate_low.to_string(formatters={
    'win_rate': '{:,.2%}'.format
})
'''
win_rate_low.win_rate = win_rate_low.win_rate * 100
win_rate_low.index += 1
win_rate_low['advantage'] = win_rate_low.win_rate - 50

print("In ", size_low, " recorded AOE2 low elo ( < 1000 ) ranked 1v1 games, the win rate of the civs are as follows: \n")


print(win_rate_low)


```

```python
#analysis for middle elo games


win_rate_mid = win_rate_table_mid['civ.win.name'].value_counts()/(win_rate_table_mid['civ.win.name'].value_counts()+win_rate_table_mid['civ.lose.name'].value_counts())
win_rate_mid = win_rate_mid.sort_values(ascending = False).reset_index().rename_axis("Rank", axis='columns').rename(columns={'index':'civs',0:'win_rate'})
'''
win_rate_mid = win_rate_mid.to_string(formatters={
    'win_rate': '{:,.2%}'.format
})
'''

win_rate_mid.win_rate = win_rate_mid.win_rate * 100
win_rate_mid.index += 1
win_rate_mid['advantage'] = win_rate_mid.win_rate - 50

print("In ", size_mid, " recorded AOE2 middle elo (1000-1300) ranked 1v1 games, the win rate of the civs are as follows: \n")


print(win_rate_mid)

```

```python
#analysis for high elo games


win_rate_high = win_rate_table_high['civ.win.name'].value_counts()/(win_rate_table_high['civ.win.name'].value_counts()+win_rate_table_high['civ.lose.name'].value_counts())
win_rate_high = win_rate_high.sort_values(ascending = False).reset_index().rename_axis("Rank", axis='columns').rename(columns={'index':'civs',0:'win_rate'})
'''win_rate_high = win_rate_high.to_string(formatters={
    'win_rate': '{:,.2%}'.format
})
'''
win_rate_high.win_rate = win_rate_high.win_rate * 100
win_rate_high.index += 1
win_rate_high['advantage'] = win_rate_high.win_rate - 50

print("In ", size_high, " recorded AOE2 high elo (>= 1300) ranked 1v1 games, the win rate of the civs are as follows: \n")
print(win_rate_high)

```

From the above comparisons, we can define a good civ to be a civ with advantage > 2% in any of the sub-instances. Ans we can conclude the following:

1. Vikings is an excellent civ for beginners with an advantage of 4.7%.
2. Goths is an interesting civ, it is advantagous in low and high elo games but disadvantagous in mid elo games. An explanation can be: Goths has a no-brainer strategy (spamming infantries), which is hard for beginners to play against. But with more experiences, mid elo players have no difficulties playing against Goths. But in high elo games, players have enough techniques to exploit the Goths extremely early and extremely late game advantages. 
3. Bulgarians is a very good civ for low and mid elo players but a poor civ in in high elo games. 
4. Incas is similar to Goths: good in low and high elo games but still ok in mid elo games. 
5. Franks has a steady 2.7% advantage in all three sub-instances. Hence we can say the Franks is an 'all-rounded' civ. 
6. Magyars' advantage decreases with the elo. Hence Magyars is a good civ for non-pro players.
7. Lithuanians is the second-best all-rounded civ. 
8. Berbers, huns and Teutons are no beginner civs but are good for experienced players. 
9. Italian is surprisingly good in mid-elo games. 
10. Celts' advantage increases with the elo.

2. Franks and Lithuanians are all-rounded civs and are good in any elo games. 


Furthermore, note that most of the good civs above are 'calvary' civs and there is literally no good 'archery' civs. We can say that in 1v1 ranked games, a calvary civ is more advantagous than an archer civ in general. 

```python
combine = win_rate_high.merge(win_rate_mid, left_on='civs', right_on='civs',
...           suffixes=('_high', '_mid'))

change = combine.win_rate_high-combine.win_rate_mid

#combine['win_rate_high'] = win_rate_high*100
#combine['win_rate_mid'] = win_rate_mid*100
combine['change_in_percentage'] = change*100

print('The comparison of win rates on high (>1300) and middle (1000-1300) elo games is as follows: ')

print(combine.sort_values(by = 'change_in_percentage',ascending = False))
```

This proves a saying that the "Chinese civilization is good for pro but poor for noobs".

```python
combine = win_rate_mid.merge(win_rate_low, left_on='civs', right_on='civs',
...           suffixes=('_mid', '_low'))

change = combine.win_rate_mid-combine.win_rate_low

combine['change_in_percentage'] = change*100

print('The comparison of win rates on mid (1000-1300) and low (<1000) elo games is as follows: ')
print(combine.sort_values(by = 'change_in_percentage',ascending = False))
```
