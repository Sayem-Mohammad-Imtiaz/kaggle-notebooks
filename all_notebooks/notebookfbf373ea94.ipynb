```python
#nested function içiçe fonksiyon
def square():
   
    def add():
       
        x = 2
        y = 3
        z = x + y
        return z
    return add()**2
print(square()) 

#çıktı 25
```

```python
def f(a, b = 1, c = 2):
    y = a + b + c
    return y
print(f(5))  #a=5

print(f(5,4,3)) #a=5 b=4 c=3

#çıktı 8
      #12
```

```python
def f(*args):         #args listelerde kullanılır
    for i in args: 
        print(i)
f(1)
print("")
f(1,2,3,4)

def f(**kwargs):      #kwargs dictionary de kullanılır
    
    for key, value in kwargs.items():              
        print(key, " ", value)
f(country = 'spain', capital = 'madrid', population = 123456)
```

```python
# lambda function fonksiyon yazmanın kısa ve kolay yolu
def square(x):
    return x**2 #xkare
print(square(5)) #x=5

square = lambda x: x**2     # xkare
print(square(4))            #x=4
tot = lambda x,y,z: x+y+z   # x,y,z topla
print(tot(1,2,3))           #x=1 y=2 z=3

#çıktı  16
        #6
```
