```python
wine_data=pd.read_csv("../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv")
wine_data.head()
```

```python
taste=[]
quality=wine_data['quality'].values
for num in quality:
    if num<6:
        taste.append("bad")
    elif num>6:
        taste.append("good")
    else: 
        taste.append("normal")
wine_data['taste'] = pd.DataFrame(taste)
wine_data=wine_data.drop(columns=['quality'])
wine_data.head(1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
wine_data['taste']=le.fit_transform(wine_data['taste'])
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range = (0,2))
scaler.fit_transform(wine_data)
```

```python

#split X and Y
Y=wine_data['taste']
X=wine_data.drop(columns=['taste'])
#split the test and train dataset
from sklearn.model_selection import train_test_split
x_trg,x_test,y_trg,y_test=train_test_split(X,Y, test_size=0.2)


print(x_trg.shape)
print(y_trg.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
import time
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
start_time = time.time()

print("--- %s seconds ---" % (time.time() - start_time))
model.fit(x_trg,y_trg)
pred=model.predict(x_test)
pred
end_time = time.time()

print("--- %s seconds ---" % (end_time-start_time))
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(pred,y_test))
print(accuracy_score(pred,y_test))
print(classification_report(pred,y_test))
```

```python
from sklearn.linear_model import LogisticRegression
modelLR=LogisticRegression(solver='liblinear')
modelLR.fit(x_trg,y_trg)
predLR=modelLR.predict(x_test)
predLR
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(predLR,y_test))
print(accuracy_score(predLR,y_test))
print(classification_report(predLR,y_test))
```

```python
from sklearn.svm import SVC
modelSV=SVC()
modelSV.fit(x_trg,y_trg)
predSV=modelSV.predict(x_test)
predSV
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print(confusion_matrix(predSV,y_test))
print(accuracy_score(predSV,y_test))
print(classification_report(predSV,y_test))
```
