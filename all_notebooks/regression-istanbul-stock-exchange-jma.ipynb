```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from sklearn.linear_model import LinearRegression

%matplotlib inline
```

```python
df = pd.read_csv("../input/istanbul-stock-exchange/istanbul_stock_exchange.csv")
df = df.set_index("date")
df.head()
```

```python
sns.pairplot(df.iloc[:, 1:])
```

```python
df.info()
```

```python
plt.figure(figsize=(10, 5))
df.plot.line()
```

```python
df.columns
```

```python
n_train = 525
```

```python
training = df.iloc[:n_train, :]
training.tail()
```

```python
testing = df.iloc[n_train:, :]
testing.head(10)
```

```python
X_train = training.iloc[:, 1:]
y_train = training.iloc[:, 0]
```

```python
X_train.head()
```

```python
y_train
```

```python
X_train = X_train.values
```

```python
X_test = testing.iloc[:, 1:].values
y_test = testing.iloc[:, 0].values
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
lr.intercept_, lr.coef_
```

```python
lr.score(X_test, y_test)
```

```python
y_test_pred = lr.predict(X_test)
```

```python
pd.DataFrame({"true": y_test, "predict": y_test_pred})
```

```python
residuals = y_test_pred - y_test
plt.scatter(range(len(residuals)), residuals)
plt.xticks(range(len(residuals)), 
           df.iloc[n_train:, ].index, rotation = 90)
plt.tight_layout()
plt.hlines([0], xmin=0, xmax=10, linestyles = "--")
plt.xlabel("Date")
plt.ylabel("Residual")
```

```python
residuals
```

```python
def rmse(y_true, y_pred):
    return np.sqrt(np.mean((y_true - y_pred) ** 2))
```

```python
rmse(y_test, y_test_pred)
```
