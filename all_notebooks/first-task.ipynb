```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../input/have-you-ever-eaten-these-foods/responses.csv")

df = df.drop(["Race", "Home location", "Gender"], axis="columns")

pd.set_option('display.max_columns', None)
```

```python
df.info()
```

```python
df.head(27)
```

```python
# I'll assume that the NaN values are false and replacing them in the dataframe
df = df.fillna("False")
```

```python
df.isnull().sum()
```

```python
df.rename(columns={"Age in years": "age"},inplace=True)
df["true"] = df.apply(lambda row: sum(row[1:20]==1) ,axis=1)
```

```python
df.head()
```

```python
mean_count_by_age = df.groupby("age")["true"].sum()/df.groupby("age")["age"].count()
mean_count_by_age = mean_count_by_age.to_frame()
mean_count_by_age.rename(columns = {0 : "food"}, inplace = True)
mean_count_by_age = mean_count_by_age.reset_index()
```

```python
mean_count_by_age
```

```python
plt.style.use('ggplot')
sns.regplot(data=mean_count_by_age, x="age", y="food", scatter_kws={'alpha':0.3})
```

The data seems to suggest that there is a slight tendency to consume more diverse food as you grow older.
