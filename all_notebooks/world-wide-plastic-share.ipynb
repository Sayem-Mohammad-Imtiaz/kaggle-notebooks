```python
import numpy as np 
import pandas as pd
from matplotlib import pyplot as  plt
import seaborn as sns
plt.style.use('seaborn')
%matplotlib inline
```

```python
df = pd.read_csv('../input/plastic-input-into-worlds-oceans/plastics-top-rivers.csv')
df.shape
```

```python
df.info()
```

```python
df.head()
```

# **Countries frequency**

```python
sns.countplot(y=df['Country'])
plt.xscale('log')
```

# **Which River from which Country?**

```python
country_rivers = df.groupby('Country')
```

```python
i = 1
for country in df['Country'].unique():
    print(f"Country: {country}")
    
    print("It's Rivers:")
    rivers =  country_rivers.get_group(country)['Entity'].values
    for river in rivers:
        print(f"{i}:{river}")
        i += 1
    i = 1
    
    print("\n")
```

# **Top 5 Rivers with highest plastic Shares** 

```python
top5_highest = df.sort_values(by='sharePlastics',ascending=False)[:5]
top5_highest
```

```python
plt.figure(figsize=(7,7))
sns.barplot(data=top5_highest, x="sharePlastics", y="Entity")
plt.legend(loc='center right')
plt.title("Top 5 Rivers with highest plastic Shares")
plt.show()
```

# **Top 5 Rivers with Least plastic Shares** 

```python
top5_least = df.sort_values(by='sharePlastics')[:5]
top5_least
```

```python
plt.figure(figsize=(7,7))
sns.barplot(data=top5_least, x="sharePlastics", y="Entity")
plt.title("Top 5 Rivers with Least plastic Shares")
plt.show()
```

# **Total Plastic Share of each Country**

```python
i = 1
sharePlastics = []
for country in df['Country'].unique():
    sharePlastics.append(country_rivers.get_group(country)['sharePlastics'].sum())
```

```python
country_plastic_share = pd.DataFrame({'Country':df['Country'].unique(),'Plastic Share':sharePlastics})
country_plastic_share
```

```python
sns.barplot(data=country_plastic_share,y='Country',x='Plastic Share')
plt.xscale('log')
plt.title("Country's Plastic Share")
plt.show()
```

# **Total 5 Countries with highest plastic shares**

```python
top5CH = country_plastic_share.sort_values(by='Plastic Share',ascending=False).head()
top5CH
```

```python
plt.figure(figsize=(7,7))
sns.barplot(data=top5CH, x="Plastic Share", y="Country")
plt.legend(loc='center right')
plt.title("Top 5 Countries with highest plastic Shares")
plt.show()
```

# **Total 5 Countries with least plastic shares**

```python
top5CL = country_plastic_share.sort_values(by='Plastic Share',ascending=False).tail()
top5CL
```

```python
plt.figure(figsize=(7,7))
sns.barplot(data=top5CL, x="Plastic Share", y="Country")
plt.legend(loc='center right')
plt.title("Top 5 Countries with Least plastic Shares")
plt.show()
```
