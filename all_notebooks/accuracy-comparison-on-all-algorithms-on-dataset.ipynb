```python
# The notebbok displays the comparison of accuracy between all the machine learning algorithms namely logistic regression,
# decision tree classifier, GaussianNB, Neighbors, etc.
# here all algorithms are added to the empty list named models, and as a result we try to find out the mean and standard
# deviation of the list.


import pandas as pd 
from sklearn import model_selection
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

df=pd.read_csv("../input/diabetes-prediction/Diabetespred.csv")

X=df.loc[:,'Pregnancies':'Age']
Y=df['Outcome']

models=[]
models.append(('LDA',LinearDiscriminantAnalysis()))
models.append(('KNN',KNeighborsClassifier()))
models.append(('DTC',DecisionTreeClassifier()))
models.append(('GNB',GaussianNB()))
models.append(('SVM',SVC()))

trainX,testX,trainY,testY=model_selection.train_test_split(X,Y,random_state=7,test_size=0.3)

for name,model in models:
    kfold=model_selection.KFold(n_splits=10,random_state=7)
    results=model_selection.cross_val_score(model,trainX,trainY)
    print("{},{},{}".format(name,results.mean(),results.std()))
```
