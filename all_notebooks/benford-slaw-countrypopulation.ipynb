```python
def getfistdigit(x):
    return int(str(x)[0])
```

```python
df['Pop first digit'] = (df['Population'].apply(getfistdigit))
df.head()
```

```python
import collections
freq= collections.Counter(df['Pop first digit'])
```

```python
freq
```

```python
sumi=0
perfreq={}
xvalues=[]
yvalues=[]

for i in freq.keys():
    sumi=sumi+freq[i]
for i in sorted (freq.keys()):
    perfreq[i]={}
    xvalues.append(i)
    perfreq[i]['count'] = freq[i]
    perfreq[i]['per'] = freq[i]/sumi*100
    yvalues.append((perfreq[i]['per']))
print(freq,perfreq)
```

```python
print(xvalues,yvalues)
bx=[1,2,3,4,5,6,7,8,9]
by=[30.1,17.6,12.5,9.7,7.9,6.7,5.8,5.1,4.6]
print(bx,by)
```
