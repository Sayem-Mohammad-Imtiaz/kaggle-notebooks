```python
def ex2(A,k):
    N = A.shape[0] #A is a N xN
    B = np.zeros((N,N)) 
    
    for i in range(N):      #row index
        for j in range(N):  #column index
                
            if (i - j > k):
                B[i,j]=0     
            else:
                B[i,-j-1]=A[i,j]
     
    return B
    
```

```python
A=np.ones((5,5))
```

```python
ex2(A,2)
```

```python
 def ex22(A,k): 
    N = A.shape[0] #A is a N xN
    B = np.zeros((N,N)) 
    
    for i in range(N):      #row index
        for j in range(N):  #column index
                
            if (i+j > N+k-1):
                B[i,j]=0   
            else:
                B[i,j]=A[i,j]
     
    return B
    
```
