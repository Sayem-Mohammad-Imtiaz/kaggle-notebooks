```python

values = list(mma["country"].value_counts().values)
index = list(mma["country"].value_counts().index)
```

```python
plt.figure(figsize=(15,15))
sns.barplot(x = index[:5], y = values[:5])
plt.xlabel('Countries')
plt.ylabel('Fighters')
plt.title('Countries of Fighters ')
```

```python

data = [ dict(
        type = 'choropleth',
        locations = index,
        locationmode = 'country names',
        z = np.log10(values),
        text = index,

        
        reversescale =True,
       
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            autotick =False,
            tickprefix = '',
            title = 'Countries Of Fighters'),
      ) ]

layout = dict(
    title = 'Countries Of Fighters',
    geo = dict(
        showframe = False,
        showcoastlines = True,
        projection = dict(
            type = 'Mercator'
        )
    )
)

fig = dict( data=data, layout=layout )
py.iplot( fig, validate=False)
```
