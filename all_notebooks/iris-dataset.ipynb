# Applying KNN Algorithm

   1. assigning target y in an array out of the full data frame (by .values)

   2. assigning data in an array out of the full data frame without the target (by .values)

```python
y = df_noID['Species'].values   
X = df_noID.drop('Species', axis=1).values 
```

  - splitting the data to training data and test data with ratio 0.7 to 0.3

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42, stratify = y)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

- training the model

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train, y_train)
```

- checking model accuracy

```python
print(knn.score(X_test, y_test))
```

 - iterating number of neighbors to check best k value

```python
neighbors = np.arange(1, 30)
score_iter = []
score = []
for k in neighbors:
    knn_iter = KNeighborsClassifier(n_neighbors=k)
    knn_iter.fit(X_train, y_train)
    score_iter= knn_iter.score(X_test, y_test)
    score.append(score_iter)
```

```python
plt.plot(neighbors, score)
plt.title('Number of neighbors vs. score')
plt.xlabel('Number of neighbors')
plt.ylabel('Model score')
plt.show()
```
