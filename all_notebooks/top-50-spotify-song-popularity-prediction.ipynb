# Importing all Libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from sklearn import metrics
from sklearn.metrics import confusion_matrix, classification_report
from xgboost import XGBRegressor
```

# Reading File

```python
Data_Spotify = pd.read_csv("../input/top50spotify2019/top50.csv",encoding='ISO-8859-1')
```

# Remving Null with 0

```python
Data_Spotify.isnull().sum()
Data_Spotify.fillna(0)
```

# Making the column names shorter for easy reading

```python
Data_Spotify.columns

Data_Spotify = Data_Spotify.rename(columns = 
                    {'Track.Name' : 'Track', 
                     'Artist.Name' : 'Artist', 
                     'Genre': 'Genre', 
                     'Beats.Per.Minute' : 'BPM',
                     'Energy' : 'Energy', 
                     'Danceability' : 'Dance', 
                     'Loudness..dB..' : 'Loud', 
                     'Liveness' : 'Liveness', 
                     'Valence.' : 'Valence',
                     'Length.' : 'Length', 
                     'Acousticness..' : 'Acoustic', 
                     'Speechiness.' : 'Speech', 
                     'Popularity' : 'Popularity'}, inplace = False)
```

# Parameters with impact on Popularity
Loud and Acoustic have less impact on Popularity. We can drop them 



# Grouping by Genre

```python
Data_Spot = Data_Spotify.groupby('Genre')['BPM','Energy','Dance','Liveness','Valence','Length','Speech','Popularity'].mean()
```

```python
Data_Spot
```

# Flattening the Data

```python
Flatten_Spot = Data_Spot.reset_index()
```

# Regression being used to train model on Mean data

```python
from sklearn.linear_model import LinearRegression

#Linear regression, first create test and train dataset
X=Flatten_Spot.loc[:,['BPM','Energy','Dance','Liveness','Valence','Length','Speech']].values
Y=Flatten_Spot.loc[:,'Popularity'].values

regressor = LinearRegression()
regressor.fit(X, Y)
print(regressor.intercept_)
print(regressor.coef_)
```

# Testing the model on Original Data

```python
X_test = Data_Spotify.loc[:,['BPM','Energy','Dance','Liveness','Valence','Length','Speech']].values
Y_test = Data_Spotify.loc[:,'Popularity'].values

Y_pred = regressor.predict(X_test)

ActVPred = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})
print(ActVPred)

#Checking the accuracy of Linear Regression
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, Y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```

# Comparing accuracy to see if there is overfitment

```python
Y_pred = regressor.predict(X)

ActVPred = pd.DataFrame({'Actual': Y, 'Predicted': Y_pred})
print(ActVPred)

#Checking the accuracy of Linear Regression
print('Mean Absolute Error:', metrics.mean_absolute_error(Y, Y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(Y, Y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y, Y_pred)))
```

```python
model = XGBRegressor()
model.fit(X_test, Y_test)
Y_pred = model.predict(X)

ActVPred = pd.DataFrame({'Actual': Y, 'Predicted': Y_pred})
print(ActVPred)

#Checking the accuracy of Linear Regression
print('Mean Absolute Error:', metrics.mean_absolute_error(Y, Y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(Y, Y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y, Y_pred)))
```
