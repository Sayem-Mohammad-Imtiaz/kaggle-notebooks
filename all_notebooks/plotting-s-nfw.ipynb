```python
# Creating list of m's and c's

M = [i for i in data['M']]
C = [c for c in data['C']]
```

```python
# Defining constant 'e'

e = float(4.094*(10**(-36)))
```

```python
# Defining SNFW function

def SNFW(m , c):
    cuberoot_value = (float(m))**(-1/3) 
    arctan_value = np.arctan (math.sqrt(c*c-1))
    log_value = math.log((c/2),10)
    S= ((1+1/c) * (1+1/c) * cuberoot_value * arctan_value * ((c*c-1)**(-1/2)))
    s_nfw = (log_value + ((1+1/c) * (1+1/c) * e * cuberoot_value * arctan_value * ((c*c-1)**(-1/2))))
    
    return(s_nfw)
```

```python
# Creating S_NFW output

S_NFW = [SNFW(m , c) for (m,c) in zip(M,C)]
```
