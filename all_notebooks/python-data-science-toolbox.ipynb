# Scope (Tanımlanan veriable'nin çeşidi
# global:En üstte tanımlanır
# local: Kodun içinde tanımlanır
# built in scope: Daha önceden python tarafından tasarlanmış inşa edici

```python
x = 2 # Global Değişkendir
def f():
    x = 3 # Yerel Değişkendir
    return x
print(x)      # x = 2 global scope
print(f())    # x = 3 local scope
```

```python
x = 5 # Global Değer
def f():
    y = 2*x        # Tanımlanan local değer
    return y
print(f())         
# İlk baş local aranır daha sonra global aranıyor. Eğer ikisi de olmazsa buitins scope aranır
```

1. # NESTED FUNCTION( İç içe fonksiyonlar)

```python
def square():
    """ Girilen sayının karesini döndürecek """
    def add():
        x = 2 #local
        y = 3  #local
        z = x + y
        return z
    return add()**2
print(square())   
```

# DEFAULT ve FLEXIBLE ARGUMENTS

def f(a, b=1): fonksiyonunda b değer alabilir eğer almazsa 1 olarak default olarak çalışır (default)

# Flexible
def f(*args): İstenilen kadar argüment alabilir

```python

def f(a, b = 1, c = 2):
    y = a + b + c
    return y
print(f(5)) # sadece a' nın değerini vermiş olduk vermesek bile tanımlı olan default değerleri alır
print(f(5,4,3)) # Burda da değerleri biz veriyoruz default değerleri almıyor(sırasıyla)
```

```python
def f(*args): # alacağı argüment belirlemiyor fazla sayıda alabilir
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4)
# **kwargs methodu dictionary de kullanılır (keyword ve value değer olanlar için)
def f(**kwargs):
    for key, value in kwargs.items(): # items ile dictionary değerlerini görebiliyoruz      
        print(key, " ", value)
f(country = 'spain', capital = 'madrid', population = 123456)
```

# LAMBDA FUNCTION ( fonksiyon yazmanın kolay loyu)

```python
square = lambda x: x**2  # lambda ile kolay yol
print(square(4))
tot = lambda x,y,z: x+y+z   
print(tot(1,2,3)) # verilen değerin toplamını verir
```

# ANONYMOUS FUNCTİON

```python
number_list = [1,2,3]
y = map(lambda x:x**2,number_list) # değerleri sırasıyla işleyip yazdırıyor
print(list(y))
```

```python
def sum(a,b):

    def sub(a,b):

        return a-b # 5 - 1 

    return sub(a,b) + 2  # 4=(sub) + 2 

print(sum(5,1))
```

# Zip Metodu ( Birleştirme)
**zip(): zip lists**

```python
list1 = [1,2,3,4]
list2 = [5,6,7,8]
z = zip(list1,list2) # zip birleştirmesi
print(z) # obje olarak tanımlanıyor
z_list = list(z)
print(z_list)
```

```python
un_zip = zip(*z_list)
un_list1,un_list2 = list(un_zip) # unzip returns
print(un_list1)
print(un_list2)
print(type(un_list2))
```

# LIST COMPREHENSİON

```python
num1 = [1,2,3]
num2 = [i + 1 for i in num1 ] #LIST COMPREHENSİON
print(num2)
```

```python
# Conditionals on iterable
num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1] 
''' num1 listesi içindekilerini for döngüsüne al, eğer i=10 ise i'nin karesini al ve yazdır
    eğer 10 değilse alınan değeri 5 den çıkar ve yazdır, sonra alınan değeri 5 ekle yazdır   '''
print(num2)
```
