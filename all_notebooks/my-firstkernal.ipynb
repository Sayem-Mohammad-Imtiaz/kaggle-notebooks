```python
# SCOPE 
# global:defined in main body in script
# local: defined in a function 
# built-in:names in predefined built in scope module such as print,len aslında reseved fonksiyonlar.

#örnek
# guess print what
x= 2 # bu golabal dir 

def f(): # f metodu
    x=3
    return x
print(x) # x = 2 global scope, ana script içinde tanımlanmış bir değişkendir,main body metod içindeki değeri bilmez.
print(f()) # x = 3 local scope, fonksiyon içinde tanımlanmıştır ve çağrılınca döndrülür. 
```

```python
# default arguments
def f (a,b =1,c =2):
    y = a+b+c
    return y
print(f(5,))
# what if we want to change default arguments
print(f(5,4,3)) # artık b ve c için yeni değerler gönderiyoruz ve 1 ve 2 değerlerii yok sayılıyor ve yeni gönderdiğim değerler kullanılarak hesaplama yapılıyor.
```

```python
# flexble arguments * args
def f(*args):
    for i in args:
        print(i)
print(f(1))
print("")
f(1,2,3,4) # sırayla veridğimiz değerlerin hepsini alır ve istediğimiz gibi kullanabiliriz.özellikle listeler 

# flexible arguments  **kwargs that is dictionary -- yani key ve value değerleri vardır 
def f(**kwargs):
    for key,value in kwargs.items():
        print(key," ",value)

f(country='spain',capital='madrid',population=123456) # bu şekilde dictionary biçimde değerler gönderebiliriz.
# burda key= country,capital,population -- valur değerleri spain, madrid ve 123456 değerleridir


```

```python
# LAMBDA FUNCTION-- faster way to write function
def square(x):
    return x**2
print(square(5))

# lambda function
square = lambda x: x**2  # x is the name of arguments(değişken)
print(square(4))

tot = lambda x,y,z: x+y+z  # x y z names of arguments
print(tot(1,2,3))

```

```python
#ANONYMOUS FUNCTION
#Like lambda fuction but it can take one than more items
#map(func,seq)  fonktion ile kullanılıyor
number_list = [1,2,3]
y = map(lambda x:x**2,number_list)
print(list(y))
```

```python
#ITERATORS
# iterable is an object that can return an iterator
# example list ,string, dictionary  birer iter objesidir
#[1,4,9,]  bir for döngüsü ile içindeki değerlere erişebiliyoruz
# sample
name = "ronaldo"
it = iter(name)
print(next(it))  # print next iteration
print(*it) #print remaining iteration

```

```python
# zip example 
liste1 = [1,2,3]
liste2 = [4,5,6]
z = zip(liste1,liste2) # sıkıştırma ama pythonda birleştirmek amaçlı kullanılıyor
print(z)

z_liste = list(z)
print(z_liste)


```

```python
#unzip
un_zip = zip(*z_liste)
un_liste1,un_liste2 = list(un_zip) # unzip returns tuple
print(un_liste1)
print(un_liste2)
print(type(un_liste2))
print(type(list(un_liste1)))
```

```python
# LIST COMPREHENSIONS 
# çok önemli bir konu. çok iyi öğrenilmesi gerekiyor.
# we use list comprehensions for data analysis often
# collapse for loops for building  lists into s singele line
# example: num1=[1,23] and we want to make it num2=[2,3,4] . this can be done with for loop.Howeverit is unnecessarily
#long. we can make it one line code that is list comprehensions.
#example 
num1 = [1,2,3]
num2 = [i +1 for i in num1]
print(num2)
```

```python
# conditional iterable
num1=[5,10,15]
num2 = [i**2 if i == 10 else i-5 if i<7 else i+5 for i in num1]
print(num2)
```

```python
num1 = [1,2,3,4]

num2 = [i**2 if i % 2 == 0 else i-i  for i in num1]

print(num2)
```
