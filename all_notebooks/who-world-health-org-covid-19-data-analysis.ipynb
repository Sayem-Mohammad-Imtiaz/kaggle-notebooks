<b>In both datasets, <i>'Deaths - newly reported in last 24 hours'</i> column contain only null values. Hence, it is better to drop that column.</b>

```python
who_covid_data28 = who_covid_data28.dropna()
who_covid_data31 = who_covid_data31.dropna()
```

```python
def replace_value(data):
    old_country =['United States of America', 'Russian Federation', 'The United Kingdom' , 'Iran (Islamic Republic of)' , 'Czechia' ,
                  'occupied Palestinian territory, including east Jerusalem' , 
                 'Republic of Moldova' , 'Viet Nam' , 'Bolivia (Plurinational State of)' , 'Venezuela (Bolivarian Republic of)' , 
                  'Republic of Korea' , 'Kosovo[1]',
                  'Syrian Arab Republic', 'United Republic of Tanzania']
    new_country = ['United States', 'Russian', 'United Kingdom', 'Iran', 'Czech Republic', 'Palestine, State of',
                   'Moldova', 'Vietnam', 'Bolivia', 'Venezuela', 'Korea, Republic of', 'Kosovo', 'Syria', 'Tanzania']
    for i in range(0,len(old_country)):
        data['Country']= data['Country'].replace(old_country[i], new_country[i])
    return data

who_covid_data28 = replace_value(who_covid_data28)
who_covid_data31 = replace_value(who_covid_data31)
```

```python
covid_data28 = pd.DataFrame()
covid_data28['Country'] = who_covid_data28['Country']
covid_data28['Continent'] = who_covid_data28['Continent']
covid_data28['Recovered cumulative total'] = who_covid_data28['Cases - cumulative total'] - who_covid_data28['Deaths - cumulative total']
covid_data28['Recovered cumulative total per 100000 population'] = who_covid_data28['Cases - cumulative total per 100000 population'] - who_covid_data28['Deaths - cumulative total per 100000 population']
covid_data28['Recovered newly reported in last 7 days'] = who_covid_data28['Cases - newly reported in last 7 days'] - who_covid_data28['Deaths - newly reported in last 7 days']
covid_data28['Recovered newly reported in last 7 days per 100000 population'] = who_covid_data28['Cases - newly reported in last 7 days per 100000 population'] - who_covid_data28['Deaths - newly reported in last 7 days per 100000 population']
covid_data28['Recovered newly reported in last 24 hours'] = who_covid_data28['Cases - newly reported in last 24 hours'] - who_covid_data28['Deaths - newly reported in last 24 hours']
covid_data28
```

```python
covid_data31 = pd.DataFrame()
covid_data31['Country'] = who_covid_data31['Country']
covid_data31['Continent'] = who_covid_data31['Continent']
covid_data31['Recovered cumulative total'] = who_covid_data31['Cases - cumulative total'] - who_covid_data31['Deaths - cumulative total']
covid_data31['Recovered cumulative total per 100000 population'] = who_covid_data31['Cases - cumulative total per 100000 population'] - who_covid_data31['Deaths - cumulative total per 100000 population']
covid_data31['Recovered newly reported in last 7 days'] = who_covid_data31['Cases - newly reported in last 7 days'] - who_covid_data31['Deaths - newly reported in last 7 days']
covid_data31['Recovered newly reported in last 7 days per 100000 population'] = who_covid_data31['Cases - newly reported in last 7 days per 100000 population'] - who_covid_data31['Deaths - newly reported in last 7 days per 100000 population']
covid_data31['Recovered newly reported in last 24 hours'] = who_covid_data31['Cases - newly reported in last 24 hours'] - who_covid_data31['Deaths - newly reported in last 24 hours']
covid_data31
```

```python
who_covid_data_diff = pd.DataFrame()
who_covid_data_diff['Country'] = covid_data28['Country']
who_covid_data_diff['Continent'] = covid_data28['Continent']
columns = list(covid_data28.columns)
for i in range(2,len(columns)):
    who_covid_data_diff[columns[i]] = covid_data31[columns[i]] - covid_data28[columns[i]]  

who_covid_data_diff
```
