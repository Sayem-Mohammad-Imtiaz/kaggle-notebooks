```python
from scipy import stats
```

```python
cross1 = pd.crosstab(df['European Union Membership'],df['Level of development'])
print(cross1)
print("\n")
print("-"*10,"CHI-SQAURE TEST","-"*10)
print("\n")
chi2 = stats.chi2_contingency(cross1)
print("Chi Square value :",chi2[0])
print("P-value :",chi2[1])
print("DOF :",chi2[2])
print("Expected frequencies : \n",chi2[3])
```

```python
eu=df[df['European Union Membership']=='Member']
noneu=df[df['European Union Membership']=='Not Member']

def mann_whitney_u_test(distribution_1, distribution_2):
    """
    Perform the Mann-Whitney U Test, comparing two different distributions.
    Args:
       distribution_1: List. 
       distribution_2: List.
    Outputs:
        u_statistic: Float. U statisitic for the test.
        p_value: Float.
    """
    u_statistic, p_value = stats.mannwhitneyu(distribution_1, distribution_2)
    return u_statistic, p_value
```

```python
mann_whitney_u_test(list(eu['Women Entrepreneurship Index']), list(noneu['Women Entrepreneurship Index']))
```

```python
mann_whitney_u_test(list(eu['Entrepreneurship Index']), list(noneu['Entrepreneurship Index']))
```
