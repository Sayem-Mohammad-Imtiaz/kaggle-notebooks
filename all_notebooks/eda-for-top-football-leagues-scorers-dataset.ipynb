### Simple Small EDA For Top Soccer Scorer League to understand the data set and get some insights

```python
import numpy as np
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv("../input/top-football-leagues-scorers/Data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df =pd.DataFrame(df)
```

```python
df.describe()
```

```python
# Luis Suarez Goals 
df[['Goals', 'Year']][df['Player Names'] == 'Luis Suarez']
```

```python
# Sdf.to_excel("Covid-19_.xlsx")
plt.scatter(df['xG'].iloc[:10],df['Player Names'].iloc[:10],color='r')
```

```python
plt.Circle(df['Club'])
```

```python
# Get top scorers that play in Barcelona 
df[['Player Names','Goals','Year']][df['Club']=='(BAR)']
```

```python
# Compare Between Lionel Messi and Cristiano Ronaldo Goals
players = df[['Player Names','Goals','Year']][(df['Player Names']=='Cristiano Ronaldo') | (df['Player Names']=='Lionel Messi')]
```

```python
players
```

```python
import seaborn as sns
```

```python
# LETS Plot The comparison
plt.figure(figsize=(10, 7))
sns.barplot(x='Player Names', y='Goals', hue='Year', data=players)
```
