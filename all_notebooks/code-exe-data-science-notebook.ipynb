```python
train_df = pd.read_csv("/kaggle/input/titanic/train.csv")
train_df
```

```python
mask = train_df["Age"] > 18.0
children_df = train_df[mask]
children_df["Survived"].sum()/len(children_df)
```

```python
mask1 = train_df["Sex"] == "female"
women_df = train_df[mask1 & mask]
women_df["Survived"].sum()/len(women_df)
```

```python
mask2 = train_df["Sex"] == "male"
men_df = train_df[mask & mask2]
men_df["Survived"].sum()/len(men_df)
```

```python
plt.hist(train_df["Age"])
plt.xlabel("Age in years")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(train_df["Fare"], train_df["Age"])
plt.ylabel("Age in years")
plt.xlabel("Fare in pounds")
plt.show()
```

```python
plt.boxplot(train_df.Fare)
plt.show()
```

```python
X = train_df["Age"].values.reshape(-1, 1)
Y = train_df["Fare"].values.reshape(-1, 1)

model = LinearRegression()
model.fit(X, Y)

Y_predictions = model.predict(X)
```

```python
p1 = plt.subplot(111)
plt.scatter(X, Y, color='green')
plt.xlabel("Age")
plt.ylabel("Fare")
p1.plot(X, Y_predictions, color='red')
plt.show()
```
