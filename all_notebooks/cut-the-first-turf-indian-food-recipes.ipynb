Indian food recipes Data visualization

This kernel consist of Data visualization on the food recipies on the Archana's Kitchen website. Archana's story from working as an Engineer to becoming one of the most successful food bloggers in India motivated me to work on this kernal.




```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import plotly.express as px
import plotly.graph_objects as go

from wordcloud import WordCloud,STOPWORDS
```

```python
df = pd.read_csv("../input/6000-indian-food-recipes-dataset/IndianFoodDatasetCSV.csv",index_col=0)
```

```python
df.head()
```

Remove unneeded features

```python
df.drop(["Instructions", "Ingredients",  "RecipeName","URL"], axis = 1, inplace = True)
df.head()
```

Lets check for null values

```python
sns.heatmap(df.isnull(),cbar=False,cmap='RdPu_r')   
plt.show()
```

We do not see any null values . we can start exploring the data

```python
df.describe(include=['object'])
```

```python
print(df.shape)
```

```python
Top_recipies = df.TranslatedRecipeName.value_counts().head()
Top_recipies
```

```python
df['Cuisine'].value_counts()
```

```python
sns.set(rc={'figure.figsize':(20,10)})

sns.countplot('Diet',data=df,palette='twilight_shifted')
plt.xticks(rotation=80)
plt.title('Types of Diet')
```

Majority of the recipies are vegeterian. 

Lets look for preparation,cooking and total time needed for cooking

```python
fig = plt.figure(figsize=(20,26))

ax1 = fig.add_subplot(231)
ax1=sns.violinplot(x='PrepTimeInMins',data=df, orient='v') 
ax2 = fig.add_subplot(232)
ax2=sns.violinplot(x='CookTimeInMins',data=df,orient='v')
ax3 = fig.add_subplot(233)
ax3=sns.violinplot(x='TotalTimeInMins',data=df, orient='v')
```

Top 10 courses on the website

```python
Top_10_courses=df.Course.value_counts()[:10].plot.bar(title="Top 10 Courses",color="r")
Top_10_courses.set_xlabel('Course',size=15)
Top_10_courses.set_ylabel('count',size=15)
```

Top 10 Cuisines

```python
plt.rcParams['figure.figsize'] = 14,7
df.Cuisine.value_counts().nlargest(10).plot(kind = "barh",color="limegreen")
plt.title("Top 10 cuisines")
plt.xlabel("Cuisine")
plt.show()
```

Types of Diet

Total Time required to make each type of Diet recipes

```python
sns.set(rc={'figure.figsize':(50,10)})

sns.catplot(x='TotalTimeInMins',y='Diet',kind='bar',data=df)
plt.title('Total time required for each Diet')
```

Total time required for each type of course meal

```python
sns.set(rc={'figure.figsize':(50,30)})

sns.catplot(x='TotalTimeInMins',y='Course',kind='point',data=df,color='g')
plt.title('Total time required for each Course')
```

```python
stopwords = set(STOPWORDS) 
def WC(values):
    wc = WordCloud(width = 500, height = 300, background_color ='black', stopwords = stopwords, min_font_size = 5,colormap='autumn').generate(values) 
    plt.axis('off')
    plt.figure(figsize=(20,11))
    plt.title("TranslatedRecipeName")
    plt.imshow(wc)
    plt.show()
k = ''
for x in list(df['TranslatedRecipeName']):
    k+=''.join(x.split(','))
WC(k)
```

```python
stopwords = set(STOPWORDS) 
def WC(values):
    wc = WordCloud(max_words=100,width = 500, height = 300, background_color ='white', stopwords = stopwords, min_font_size = 5,colormap='cool').generate(values) 
    plt.axis('off')
    plt.figure(figsize=(20,11))
    plt.title("TranslatedInstructions")
    plt.imshow(wc)
    plt.axis('off')
    plt.show()
k = ''
for x in list(df['TranslatedInstructions']):
    k+=''.join(x.split(','))
WC(k)
```
