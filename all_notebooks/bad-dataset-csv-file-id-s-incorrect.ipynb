### create a list of file names, read in df and see if id_code in df are files in the image directory 

```python
image_dir=r'../input/aptos-balanced-dataset/train_images/train_images'
file_list=os.listdir(image_dir)
df=pd.read_csv(r'../input/aptos-balanced-dataset/train_new.csv')
print (df.head())
print ('number of images: ', len(file_list), ' number of CSV id_codes: ', len(df['id_code']))
print('NOTICE IN MISMATCH OF NUMBER OF FILES TO NUMBER OF CSV id_codes')
errors=0
error_list=[]
for f in df['id_code']:  #id_code entries do not include the extension but it should not matter for the comparison below
    if f not in file_list:
        errors +=1
        error_list.append(f)
print ('There are ', errors, ' CSV id_codes that do not exists in the images directory')
# some example errors
print ('Example errant csv id_codes is shown below')
for i in range (5):
    print (error_list[i])
```
