```python
topSals = hit.sort_values(by = ['salary'], ascending  = False)
topSals['salary']
```

I didn't expect to see a salary column, it sparks my interest. I wonder if salary corralates to play.

I think it will be useful to look at three key stats:
* Batting Average (avg)
* On-base Plus Slugging (OPS)
* Wins Above Replacement (WAR)

```python
sns.scatterplot(x = hit['avg'], y = hit['salary'])
```

It seems like there are a large amount of players inbetween .2 and .4. The highest paid are closer to .4, which means that players are being paid somewhat equitably.

```python
sns.scatterplot(x = hit['OPS'], y = hit['salary'])
```

OPS contians onbase percentage and slugging percentage. It is supposed to tell the complete picture of a hitter. We can see that the highest paid players are at the to of the pay range for the most part.

```python
sns.scatterplot(x = hit['OPS'], y = hit['salary'], hue = hit['age']>30)
```

This accounts for age. The orange is of players over 30. This shows how older players are higher on the pay scale and younger players are not paid as much.

```python
sns.scatterplot(x = hit['war'], y = hit['salary'])
```

WAR is the new thing in baseball. It is supposed to be an all encompassing stat. We can see that there is a great range and doesn't tell us anything
