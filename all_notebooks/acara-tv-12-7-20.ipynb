```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

from jcopml.pipeline import num_pipe, cat_pipe
from jcopml.utils import save_model, load_model
from jcopml.plot import plot_missing_value
from jcopml.feature_importance import mean_score_decrease
```

# Import Data & Menjadikan Kolom Menjadi Indeks 

```python
df = pd.read_csv("../input/tv-shows-on-netflix-prime-video-hulu-and-disney/tv_shows.csv",index_col='Unnamed: 0')
df.head()
```

# Membuang Kolom Yang Sekiranya Tidak Perlu Untuk di Analis

```python
df.type.value_counts() #untuk melihat apa saja dari kolom tersebut
```

```python
df.drop(columns=["type"],inplace=True)
df.head()
```

#  Untuk Melihat Data Yang Bolong

```python
plot_missing_value(df)
```

```python
df= df[~df.IMDb.isna()]
df= df[~df.Age.isna()]
df= df[~df["Rotten Tomatoes"].isna()]
```

Noted: Untuk menghilangkan data yang bolong ,yang bertujuan umtuk mempermudah analis

```python

plot_missing_value(df)
```

noted : Akhirnya datanya jadi tidak ada yang bolong

# Cara Membikin Plot Dengan Memakai Seaborn

```python
plt.figure(figsize=(15,10))

plt.subplot(241)
sns.countplot("Netflix", data=df)

plt.subplot(242)
sns.countplot("Hulu", data=df)

plt.subplot(243)
sns.countplot("Prime Video", data=df)

plt.subplot(244)
sns.countplot("Disney+", data=df)
```

noted : Platfrom streaming mana yang sering digunakan untuk menayangkan serial tv

# Cara Membikin Plot Dengan Histogram
