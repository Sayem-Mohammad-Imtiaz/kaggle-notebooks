```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (15, 32)
plt.rcParams["font.size"] = 19

df = pd.read_csv("/kaggle/input/total-covid19-tests-performed-by-country/covid-19-tests-country.csv")

tarr = ['Armenia', 'Australia', 'Australia ', 'Austria', 'Bahrain', 'Belarus', 'Belgium', 'Brazil', 'Canada ', 'China ', 'Colombia', 'Costa Rica', 'Croatia', 'Czech Republic', 'Denmark', 'Estonia', 'Faeroe Islands', 'Finland', 'France', 'Germany', 'Hong Kong', 'Hungary', 'Iceland', 'India', 'Indonesia', 'Iran', 'Ireland', 'Israel', 'Italy', 'Japan', 'Kyrgyzstan', 'Latvia', 'Lithuania', 'Malaysia', 'Malta', 'Mexico', 'Netherlands', 'New Zealand', 'Norway', 'Pakistan', 'Palestine', 'Panama', 'Philippines', 'Poland', 'Qatar', 'Romania', 'Russia', 'Slovakia', 'Slovenia', 'South Africa', 'South Korea', 'Spain', 'Sweden', 'Switzerland', 'Taiwan', 'Thailand', 'Turkey', 'Ukraine', 'United Arab Emirates', 'United Kingdom', 'United States', 'United States ', 'Vietnam']
outs = [0.0]*len(tarr)

countrys = df['Entity']
tests = df['Total COVID-19 tests']

for k in range(0, len(countrys)):
    line_s = countrys[k].split("-")
    outs[tarr.index(line_s[0])] += tests[k]

xx, yy = [], []

for i in range(0, len(tarr)):
    tarr[i] = [tarr[i], outs[i]]

tarr.sort(key=lambda xx: xx[1], reverse=False)

for i in range(0, len(tarr)):
    xx.append(tarr[i][0])
    yy.append(tarr[i][1])
    
plt.title("Total number of COVID-19 tests by country")
plt.xlabel("Number of tests")
plt.ylabel("Countries")
plt.grid(linestyle="--", linewidth=1.618)
plt.xticks(rotation=31)
plt.plot(yy, xx, linewidth=1.618, marker="o")
plt.show()
```
