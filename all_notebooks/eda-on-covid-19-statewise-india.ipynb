```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
location_data= "../input/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv"
```

```python
covid_data=pd.read_csv(location_data)
covid_data
```

```python
#comparison for different states.
plt.figure(figsize=(20,10))
sns.barplot(y="State/UTs",x="Deaths",data=covid_data)
plt.title("death comparisons among states of india due to COVID-19")
```

```python
#recovery rate comparison for different states
plt.figure(figsize=(20,10))
sns.barplot(y="State/UTs",x="Discharge Ratio",data=covid_data)
plt.title("recvovery rate comparisons among states of india due to COVID-19")
plt.xlabel("recovery rate")
```
