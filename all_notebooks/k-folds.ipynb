```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
data = pd.read_csv('../input/heart.csv')
X = (data.iloc[:,:-1]).as_matrix()
y = (data.iloc[:,2]).as_matrix()
```

```python
y = y.reshape((-1,1))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import Imputer
from sklearn.neural_network import MLPClassifier as mlp_c
#my_pipeline = make_pipeline(Imputer(), RandomForestRegressor())
clf = mlp_c(solver='lbfgs', alpha=1e-5,hidden_layer_sizes=(100,100), random_state=1)
clf.fit(X_train, y_train) 
y_hat = clf.predict(X_test)
```

segun la evaluación que queremos obtener es comparar que tan iguales son mis dos variables, los datos de prueba y lo predecido, usando el [fowlkes_mallows_score](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.fowlkes_mallows_score.html#sklearn.metrics.fowlkes_mallows_score)

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf,y_test, y_hat, scoring='fowlkes_mallows_score')
print(scores)
```

El codigo como resultado arroja la similitud entre valores, un numero entre cero y uno (uno para iguales). 
Arroja 3 resultados producto de crear 3 k-folds y está por default, para obtener un unico resultado se saca la media.

```python
resultado=scores.mean()
print(resultado)
```
