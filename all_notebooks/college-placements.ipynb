```python
ncp = pd.get_dummies(data=cp,drop_first=True)
ncp.head()
```

```python
X = ncp.drop(columns='PlacedOrNot')
y = ncp.PlacedOrNot
```

## Model Building 

### SVC

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
# from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=22)
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
from sklearn.metrics import plot_confusion_matrix
```

```python
s = SVC(kernel='rbf',C=10000)
s.fit(X_train,y_train)
np.mean(cross_val_score(s, X_train, y_train, cv=10))
```

```python
np.mean(cross_val_score(s, X_test, y_test, cv=10))
```

```python
y_hat_svm = s.predict(X_test)
```

```python
plot_confusion_matrix(s,X_test,y_test,display_labels=['Not Placed','Placed'])
plt.show()
```

```python
print(metrics.classification_report(y_test,y_hat_svm))
```

### Logistic Regression

```python
lor = LogisticRegression(multi_class='ovr')
```

```python
lor.fit(X_train,y_train)
```

```python
y_hat_lor = lor.predict(X_test)
```

```python
lor.score(X_test,y_test)
```

```python
plot_confusion_matrix(lor,X_test,y_test)
plt.show()
```

```python
print(metrics.classification_report(y_test,y_hat_lor))
```
