```python
df_number_columns = df.select_dtypes(include = np.number)
```

```python
df_category_columns = df.select_dtypes(exclude = np.number)
```

```python
df_category_columns.columns
```

```python
data_to_encode = df_category_columns.drop(columns = ['Attrition', 'Over18'])
data_to_encode[:5]
```

```python
df_category_encoded = pd.get_dummies(data_to_encode)
```

```python
df_category_encoded[:3]
```

```python
df_number_columns.columns
```

```python
df_final = pd.concat([df_category_encoded,df_number_columns.drop(columns = ['EmployeeCount','StandardHours'])], axis=1)
```

```python
df_final[:3]
```

```python
train_X, test_X, train_y, test_y = train_test_split(df_final,df['Attrition'], test_size = 0.2)
```

```python
model = LogisticRegression()
model.fit(train_X,train_y)
```

```python
predict_train = model.predict(train_X)
predict_test = model.predict(test_X)
```

```python
print('Classification report for Train: \n', classification_report(train_y, predict_train))
```

```python
print('Classification report for Test: \n', classification_report(test_y, predict_test))
```
