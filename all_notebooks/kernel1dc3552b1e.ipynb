```python
import re
```

```python
def cln_date(txt):
    txt=re.sub('January', '.1',txt)
    txt=re.sub('February', '.2', txt)
    txt=re.sub('March', '.3', txt)
    txt=re.sub('[ ]', '', txt)
    return txt
    
```

```python
df2['Date']=df2['Date'].apply(lambda x: cln_date(x))
```
