```python
df_math['Drinks_Per_Day'] = 'NA'
df_math.loc[(df_math['Dalc'] > 0) & (df_math['Dalc'] < 2), 'Drinks_Per_Day'] = 'One'
df_math.loc[(df_math['Dalc'] > 1) & (df_math['Dalc'] <= 7), 'Drinks_Per_Day'] = 'More_Than_One'
```

```python
df_math.head()
```

```python
# Do male or female drink more than one alcoholic drink per day?
df_math[df_math.Drinks_Per_Day == 'More_Than_One'].sex.value_counts(normalize=True) #Yes
```

```python
# Are Urban students drinking more than Rural students?
df_math[df_math.Drinks_Per_Day == 'More_Than_One'].address.value_counts(normalize=True)  #Yes
```

```python
# Are students in romantic relationships drinking more than single students?
df_math[df_math.Drinks_Per_Day == 'More_Than_One'].romantic.value_counts(normalize=True)   #No
```

```python
# Average final Math grade per sex
df_math.groupby('sex')['G3'].mean()
```

```python
# Average final Portuguese grade per sex
df_port.groupby('sex')['G3'].mean()
```
