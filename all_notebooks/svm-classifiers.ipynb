```python
# Data Preprocessing
```

```python
df.drop(columns=['User ID'], axis=1, inplace=True)
```

```python
# Converting Gender into Dummy Data
gender = pd.get_dummies(df['Gender'], drop_first=True)

# Adding dummy data to data set and removing Gender column
df = pd.concat([gender, df], axis=1)
df.drop(columns=['Gender'], axis=1, inplace=True)
```

```python
# Extracting Feature
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
# Feature Scaling
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
X = ss.fit_transform(X)
```

```python
X
```

```python
# Splitting data into training and testing
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.svm import SVC
```

```python
# Using Polynomial Kernel

poly_model = SVC(kernel='poly', degree=3)
```

```python
poly_model.fit(X_train, y_train)
```

```python
poly_model.score(X_train, y_train)
```

```python
poly_model.score(X_test, y_test)
```

```python
poly_model.predict(ss.transform([[1,19,19000]]))
```

```python
y_pred_poly = poly_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cf_matrix = confusion_matrix(y_test, y_pred_poly)
sns.heatmap(cf_matrix, annot=True)
```

```python
print(classification_report(y_test, y_pred_poly))
```

```python
# Using RBF Kernel
```

```python
rbf_model = SVC(kernel='rbf').fit(X_train, y_train)
```

```python
# Score on test data

rbf_model.score(X_test, y_test)
```

```python
# Score on training data

rbf_model.score(X_train, y_train)
```

```python
rbf_model.predict(ss.transform([[0,26,43000]]))
```

```python
y_pred_rbf = rbf_model.predict(X_test)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred_rbf), annot=True)
```

```python
print(classification_report(y_test, y_pred_rbf))
```

```python
# Random Forest Classifier
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_clf = RandomForestClassifier(max_depth=4)
```

```python
rf_clf.fit(X_train, y_train)
```

```python
rf_clf.score(X_test, y_test)
```

```python
rf_clf.score(X_train, y_train)
```

```python
y_pred_rf = rf_clf.predict(X_test)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True)
```

```python
print(classification_report(y_test, y_pred_rf))
```

```python
# Decision Tree Classifiers
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_clf = DecisionTreeClassifier(max_depth=3).fit(X_train, y_train)
```

```python
dt_clf.score(X_test, y_test)
```

```python
dt_clf.score(X_train, y_train)
```

```python
y_pred_dt = dt_clf.predict(X_test)
```

```python
sns.heatmap(confusion_matrix(y_test, y_pred_dt), annot=True)
```

```python
print(classification_report(y_test, y_pred_dt))
```
