# Cleaned Temperature Changed Data

## In the following map, white lines are showing the null values in our data which we need to clean

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.isnull())
```

## Percentage of Missing Values

```python
null_var = df.isnull().sum()/df.shape[0] *100
null_var
```

## Dropping NA values

```python
df2_drop_rows = df.dropna()
```

```python
df2_drop_rows.shape
```

## Hence, we can see that our map is totally cleaned from white lines

```python
plt.figure(figsize=(15,15))
sns.heatmap(df2_drop_rows.isnull())
```

```python
df2_drop_rows.isnull().sum().sum()
```

```python
df2_drop_rows.select_dtypes(include=['int64','float64']).columns
```

## Here, we can analyze that how much difference arises after dropping the null rows

```python
sns.distplot(df['Y1965'])
```

```python
sns.distplot(df2_drop_rows['Y1965'])
```

## Here we can see that there is not much difference arises after dropping the null rows

```python
sns.distplot(df['Y1965'])
sns.distplot(df2_drop_rows['Y1965'])
```
