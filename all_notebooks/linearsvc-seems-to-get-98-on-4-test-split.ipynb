```python
data = pd.read_csv("/kaggle/input/sentimental-analysis-for-tweets/sentiment_tweets3.csv")
data.head()
```

```python
texts = data["message to examine"]
```

```python
import spacy

nlp = spacy.load("en_core_web_lg")
vectors = np.array([nlp(text).vector for text in texts])
```

```python
vectors.shape, data.shape
```

```python
labels = data.iloc[:,2]
labels
```

```python
from sklearn.svm import LinearSVC
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(vectors, labels, test_size=0.4, random_state=0)

model = LinearSVC()
model.fit(X_train, y_train)
print(X_train.shape, X_test.shape)
sum(model.predict(X_test) == y_test)/len(X_test)
```
