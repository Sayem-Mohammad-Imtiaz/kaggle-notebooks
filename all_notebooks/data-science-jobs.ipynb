# Looking At Data Science Jobs
This is a basic look at what you need to get a data sceince job at Google.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS
```

# The Data
This is a small number of examples when you start to look into it further. There are very few of the same type of job for example there are only 12 jobs that contain data science of data scientist. In a way the table below is most likly all you need to look at the Qualifications and responsibilities.

```python
gdf = pd.read_csv('../input/job_skills.csv')
gdf.shape
```

```python
datadf = gdf[gdf['Title'].str.contains('Data Science') | gdf['Title'].str.contains('Data Scientist')]
datadf['Title'].count()
gdf[gdf['Title'].str.contains('Data Science') | gdf['Title'].str.contains('Data Scientist')]
```

# Word Clouds
To not make this notebook that boring I made word clouds for Minimum Qualifications, Preferred Qualifications, Location, and Responsibilities in that order.

```python
mpl.rcParams['figure.figsize']=(8.0,7.0)
mpl.rcParams['font.size']=9                
mpl.rcParams['savefig.dpi']=100              
mpl.rcParams['figure.subplot.bottom']=.1 
#make the frame the wordcloud will be in

STOPWORDS.add('degree')
STOPWORDS.add('name')
STOPWORDS.add('dtype')
STOPWORDS.add('location')
stopwords = set(STOPWORDS)
data = datadf
#get stop words and the data

wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=100,
                          max_font_size=50, 
                          random_state=False
                         ).generate(str(data['Minimum Qualifications']))
#varibles for the wordcloud

print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
#show wordcloud
```
