Predict Rating for Board game review

```python
import pandas as pd
from sklearn import metrics
from sklearn.model_selection import train_test_split
from nltk.tokenize import RegexpTokenizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
import pickle

dataset = pd.read_csv('../input/boardgamegeek-reviews/bgg-15m-reviews.csv')
```

```python
# First Remove Empty Comment Rows and Round the Rating to the nearest Integer to avoid too many values in predicted score

dataset = dataset[dataset['comment'].notna()]
dataset['rating'] = dataset['rating'].round()

# Return Vectorizer Based On Choice

def VectorizationChoice(Type, ngramrange_start, ngramrange_end):
    if Type == 'TF-IDF':
        return TfidfVectorizer(analyzer='word',
    sublinear_tf=True,
    strip_accents='unicode',
    token_pattern=r'\w{1,}',
    ngram_range=(1, 1),
    max_features=None)
    elif Type == 'CountVectorizer':
        tokenizr = RegexpTokenizer('[a-zA-Z0-9]+')
        return CountVectorizer(stop_words='english',ngram_range = (ngramrange_start, ngramrange_end),tokenizer = tokenizr.tokenize)
    else:
        return None
```

```python
# Multinomial Naive Bayes Algorithm with monogram using TF-IDF Vectorizer

ngram_range_start = 2
ngram_range_end = 2

Vectorizer = VectorizationChoice('TF-IDF', None, None)

texts = Vectorizer.fit_transform(dataset['comment'])

train_x, test_x, train_y, test_y = train_test_split(texts, dataset['rating'], test_size=0.20, random_state=0)

MultinomialNB_Model = MultinomialNB(alpha=1)

MultinomialNB_Model.fit(train_x, train_y)

MultinomialNB_Prediction = MultinomialNB_Model.predict(test_x)

Final_Accuracy = metrics.accuracy_score(MultinomialNB_Prediction, test_y)

print('Accuracy for Multinomial Naive Bayes Algorithm with monogram using TF-IDF Vectorizer: ', end='')

print(round(Final_Accuracy, 2))
```
