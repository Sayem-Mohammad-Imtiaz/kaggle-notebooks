```python
pokemon_stats = pd.read_csv('/kaggle/input/pokemon/Pokemon.csv')
pokemon_stats.head()
```

```python
labels=np.array(['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed'])
stats=pokemon_stats.loc[200,labels].tolist()
#You need to add the first value as the last as well as the rader moves in a circular manner
stats += stats[:1]
stats
```

```python
import plotly.express as px
from math import pi
import matplotlib.pyplot as plt
```

```python
ax = plt.subplot(111, polar=True)
angles = [n / float(len(labels)) * 2 * pi for n in range(len(labels))]
angles += angles[:1]
plt.xticks(angles[:-1], labels, color='grey', size=8)
ax.plot(angles, stats, linewidth=1, linestyle='solid',color = "pink")
ax.fill(angles, stats, 'b', alpha=0.1 , color = "royalblue")
plt.yticks( color="black", size=7)


```
