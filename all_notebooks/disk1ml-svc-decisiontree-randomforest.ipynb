```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # data visualization library  
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix


df = pd.read_csv("../input/vdisk1-for-dl/disk1DL.csv")
```

```python
df.head()
```

##Checking null value

```python
df.isnull().sum()
```

##Converting string to integer

```python
from sklearn.preprocessing import LabelEncoder 
le= LabelEncoder()
```

```python
df['diskProgram'] = le.fit_transform(df['diskProgram'])
df['startTid'] = le.fit_transform(df['startTid'])
df['stoppTid'] = le.fit_transform(df['stoppTid'])
df['diskObjekt'] = le.fit_transform(df['diskObjekt'])
df
```

## Lable and feature separation

```python
y= df['lebel data']
y
```

```python
x= df.drop(['lebel data'], axis= 1)
x
```

```python
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(x.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

# **Support vactor classifier (SVC)**

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
```

```python
xtrain, xtest, ytrain, ytest= train_test_split(x, y, test_size= 0.25, random_state=1)
```

```python
svm= SVC(gamma='scale')
svm.fit(xtrain,ytrain)
svm.score(xtest, ytest)
```

**R_Squred_value**

```python
pred_svm= svm.predict(xtest)
```

```python
ac = accuracy_score(ytest,pred_svm)
print('Accuracy is: ',ac)
cm = confusion_matrix(ytest,pred_svm)
sns.heatmap(cm,annot=True,fmt="d")
```

# **Random Forest classifier**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

rfc = RandomForestClassifier()
```

```python
rfc.fit(xtrain, ytrain)
rfc.score(xtest, ytest)
```

# **RandomForestClassifier Score is : 0.97**

```python
pred = rfc.predict(xtest)

ac = accuracy_score(ytest,rfc.predict(xtest))
print('Accuracy is: ',ac)
cm = confusion_matrix(ytest,rfc.predict(xtest))
sns.heatmap(cm,annot=True,fmt="d")
print(classification_report(ytest,pred))
```
