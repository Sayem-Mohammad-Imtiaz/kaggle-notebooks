```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
```

```python
data_insta = pd.read_csv('../input/pakistans-top-25-instagram-accounts/Top Instragram.csv')
data_insta.head()
```

```python
data_insta.shape
```

```python
sns.scatterplot(data=data_insta)
```

```python
data_insta = data_insta[['Name', 'Category', 'Followers']]
data_insta
```

```python
def followers_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x:
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    
data_insta['Followers'] = data_insta['Followers'].apply(followers_to_float)
data_insta.head()
```

```python
data_insta['Category'].unique()
```

```python
# Changing category "Singer/Musician" to "Singer"
data_insta.replace("Singer/Musician", "Singer", inplace=True)
data_insta['Category'].unique()
```

```python
plt.figure(figsize=(10, 10))
dfCat = data_insta.groupby(by=['Category']).sum()
subs = list(dfCat['Followers'])
names = list(data_insta['Category'].unique())
colors = ['grey', '#ff9c52', 'red', '#e6859f', '#e69585', '#c1d685', '#87e685', '#e685dc', '#edabac', '#c8e8e5']
plt.pie(subs, labels=names, autopct='%1.1f%%', colors=colors)
plt.show()
```

```python
plt.figure(figsize=(17, 7))
plt.yticks(rotation=90)
graph = sns.barplot(x='Name', y='Followers', data=data_insta.query('Followers >= 5000000'))
```

```python
plt.figure(figsize=(10, 10))
channels = data_insta.query('Category == "Media Channel"')
subs = list(channels['Followers'])
names = list(channels['Name'])
colors = ['#baff9e', '#ff9c59', '#85dae6', '#e6859c', '#e69585', '#c1e685', '#87e685', '#e685dc', '#edabcc', '#c8e8c5']
plt.pie(subs, labels=names, autopct='%1.1f%%', colors=colors)
plt.show()
```
