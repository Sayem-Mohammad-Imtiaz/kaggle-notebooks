```python
import bs4
from bs4 import BeautifulSoup
import requests
import re
import time
from datetime import datetime

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('/kaggle/input/paralmpic-2020/Paralympic_Medal_Count.csv')
df.head()
```

```python
df.drop_duplicates(keep='first',inplace=True)
```

```python
df.info()
```

# Total number of Gold, Silver, Bronze medals

```python
count = df[['Gold','silver','bronze']].sum().reset_index().rename(columns={'index':'Medal',0:'Count'})
```

```python
plt.figure(figsize=(9,5),facecolor='black')
plt.xticks(rotation=0,fontsize=12,color='white',fontweight='bold')
plt.yticks(fontsize=12,color='black',fontweight='bold')
colors=['gold','silver','#CD7F32']
a = sns.barplot(x='Medal',y='Count',data=count,palette=colors)
for p in a.patches:
    a.annotate(format(p.get_height(),'.1f'),(p.get_x()+p.get_width()/2.,p.get_height()),ha='center',
              va='center',xytext=(0,7.2),textcoords='offset points',fontsize=14,color='white')
a.set_facecolor('black')
a.set_xlabel("Total Medals", fontsize = 20,color='pink')
a.spines['left'].set_color('black')
a.spines['top'].set_color('black')
a.spines['bottom'].set_color('white')
a.spines['right'].set_color('black')
plt.grid(False)
```

# Top 10 Teams with highest medals

```python
cols= ['Team/Npc','rank by total','total medals']
by_rank = df.sort_values('rank by total',ascending=True).reset_index().head(10)
by_rank = by_rank[cols].head(10)
```

```python
sns.set(rc={'figure.figsize':(14,7)})
plt.figure(facecolor='black')
plt.xticks(rotation=0,fontsize=16,color='white',fontweight='bold')
plt.yticks(fontsize=16,color='white',fontweight='bold')
a = sns.barplot(y='Team/Npc',x='total medals',data = by_rank,palette='spring',edgecolor='none')
for p in a.patches:
    width = p.get_width()
    plt.text(6+p.get_width(),p.get_y()+0.55*p.get_height(),'{:1.1f}'.format(width),ha='center',va='center',
             fontsize=14,color='gold',fontweight='bold')
a.set_xlabel("Number of Medals", fontsize = 20,color='pink')
plt.title("Top 10 teams with highest medals",fontsize=28,color='pink')
a.set_facecolor('black')
a.spines['left'].set_color('white')
a.spines['top'].set_color('black')
a.spines['bottom'].set_color('black')
a.spines['right'].set_color('black')
plt.grid(False)
```

# Top 10 Teams different medals

```python
by_rank = df.sort_values('rank by total',ascending=True).reset_index().head(10)
by_rank = by_rank.drop('index',axis=1)
plt.figure(figsize=(16,7))
xpos = np.arange(len(by_rank['Team/Npc'].head(10)))
plt.xticks(xpos,by_rank['Team/Npc'],rotation=90,fontweight='bold',fontsize=13)
plt.bar(xpos-0.3,'Gold',data=by_rank.head(10),color='gold',width=0.3)
plt.bar(xpos,'silver',data=by_rank.head(10),color='#C0C0C0',width=0.3,label='Silver')
plt.bar(xpos+0.3,'bronze',data=by_rank.head(10),color='#C39B77',width=0.3,label='Bronze')
plt.yticks(fontsize=12,color='black',fontweight='bold')
plt.grid(False)
plt.legend()
plt.show()
```
