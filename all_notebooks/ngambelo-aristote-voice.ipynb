These information sometimes can be useful to select features to be used for training model. So now is the time to create training logic. Data set first should be separated with training data and test data for evaluating trained model.

```python
df=df.drop(['maxdom','centroid','kurt'],axis=1)
```

```python
X=df.drop(['label'],axis=1)
y=df['label']
```

```python
y=y.map({'male':0,'female':1})
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn import tree
dtc=tree.DecisionTreeClassifier()
dtc.fit(X_train,y_train)
y_dtc=dtc.predict(X_test)
cm=confusion_matrix(y_test,y_dtc)
print(accuracy_score(y_test,y_dtc))
print(cm)
print(classification_report(y_test,y_dtc))
```

```python
sns.heatmap(cm,annot=True)
plt.xlabel('actual')
plt.ylabel('y_predict')
```

```python
from sklearn import ensemble
rf=ensemble.RandomForestClassifier()
rf.fit(X_train,y_train)
y_rf=rf.predict(X_test)
cm=confusion_matrix(y_test,y_rf)
print(accuracy_score(y_test,y_rf))
print(cm)
print(classification_report(y_test,y_rf))
```

```python
sns.heatmap(cm,annot=True)
plt.xlabel('actual')
plt.ylabel('y_predict')
```

```python
n_cols = X_train.shape[1]
y_cat = to_categorical(y)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, random_state=1)
hist = History()
model = Sequential()
model.add(Dense(1000, activation='relu', input_dim = n_cols))
model.add(Dense(700, activation='relu'))
model.add(Dense(400, activation='relu'))
model.add(Dense(400, activation='relu'))
model.add(Dense(400, activation='relu'))


model.add(Dense(2, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

train=model.fit(X_train, y_train, epochs = 50, validation_split = .1, callbacks = [hist])
```

```python
model.evaluate(X_test,y_test)
```

```python
print(train.history['accuracy'])
```

```python
print(train.history['val_accuracy']) 
```

```python
def plot_scores(train):
    accuracy=train.history['accuracy']
    val_accuracy=train.history['val_accuracy']
    epochs=range(len(accuracy))
    plt.plot(epochs,accuracy,'b',label='Score apprentissage')
    plt.plot(epochs,val_accuracy,'r',label='Score validation')
    plt.title('Scores')
    plt.legend()
    plt.show()
```

```python
plot_scores(train)
```
