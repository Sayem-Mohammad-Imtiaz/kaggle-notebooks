# Reading Data

```python
df = pd.read_csv("../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv")

# take a look at the dataset
df.head()
```

Selecting features for Regression

```python
cdf = df[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','pH','alcohol','quality']]
cdf.head(9)
```

```python
plt.scatter(cdf.alcohol, cdf.pH,  color='blue')
plt.xlabel("alcohol")
plt.ylabel("pH")
plt.show()
```

Creating train and test dataset.

```python
msk = np.random.rand(len(df)) < 0.7
train = cdf[msk]
test = cdf[~msk]
```

Train data distribution.

```python
plt.scatter(train.alcohol, train.pH,  color='blue')
plt.xlabel("alcohol")
plt.ylabel("pH")
plt.show()
```

#  Multiple Regression Model

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['alcohol','chlorides','residual sugar']])
y = np.asanyarray(train[['citric acid']])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

# Prediction

```python
y_hat= regr.predict(test[['alcohol','chlorides','residual sugar']])
x = np.asanyarray(test[['alcohol','chlorides','residual sugar']])
y = np.asanyarray(test[['citric acid']])
print("Sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```
