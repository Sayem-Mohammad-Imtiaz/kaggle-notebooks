## train, test validation

```python
x=x.values
y=y.values
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.20)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,random_state=0,test_size=0.13)
```

```python
x_train.shape
```

```python
x_val.shape
```

## Normalize

```python
from sklearn.preprocessing import StandardScaler

obje_ss=StandardScaler()

x_train_ss=obje_ss.fit_transform(x_train) 
x_val_ss=obje_ss.transform(x_val)
x_test_ss=obje_ss.transform(x_test)
```

```python
x_train_ss
```

```python
x_train_ss.shape
```

## Models

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

model_log_reg=LogisticRegression(max_iter=1000).fit(x_train_ss,y_train)
y_pred_val_log=model_log_reg.predict(x_val_ss)
print('LogisticRegression classifcation train başarı oranı %',model_log_reg.score(x_train_ss,y_train)*100)
print(confusion_matrix(y_val,y_pred_val_log))
print('LogisticRegression classifcation validation başarı oranı %',(confusion_matrix(y_val,y_pred_val_log)[0][0]+confusion_matrix(y_val,y_pred_val_log)[1][1])/len(y_val)*100)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix

model_svc=SVC().fit(x_train_ss,y_train)
y_pred_val_svc=model_svc.predict(x_val_ss)
print('svc classifcation train başarı oranı %',model_svc.score(x_train_ss,y_train)*100)
print(confusion_matrix(y_val,y_pred_val_svc)) 
print('svc classifcation validation başarı oranı %',(confusion_matrix(y_val,y_pred_val_svc)[0][0]+confusion_matrix(y_val,y_pred_val_svc)[1][1])/len(y_val)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier

model_tree=RandomForestClassifier(n_estimators=10).fit(x_train_ss,y_train)
y_pred_val_tree=model_tree.predict(x_val_ss)
print('tree classifcation train başarı oranı %',model_tree.score(x_train_ss,y_train)*100)
print(confusion_matrix(y_val,y_pred_val_tree))
print('tree classifcation validation başarı oranı %',(confusion_matrix(y_val,y_pred_val_tree)[0][0]+confusion_matrix(y_val,y_pred_val_tree)[1][1])/len(y_val)*100)
```

```python
from sklearn.neighbors import KNeighborsClassifier

model_knn=KNeighborsClassifier(n_neighbors=4,metric='minkowski').fit(x_train_ss,y_train)
y_pred_val_knn=model_knn.predict(x_val_ss)
print('KNN classifcation train başarı oranı %',model_knn.score(x_train_ss,y_train)*100)
print(confusion_matrix(y_val,y_pred_val_knn))
print('tree classifcation validation başarı oranı %',(confusion_matrix(y_val,y_pred_val_knn)[0][0]+confusion_matrix(y_val,y_pred_val_knn)[1][1])/len(y_val)*100)
```
