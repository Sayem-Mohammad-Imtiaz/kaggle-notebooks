### This project is a simple visualization of current Covid-19 confirmed cases due to the new delta variant. The visualization is shown at the district level. The marker here represents a district of a corresponding division and the pop-up has the case number.

```python
from  geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="http")
import folium
import pandas as pd
```

```python
df=pd.read_csv("/kaggle/input/covid19-confirmed-cases-due-to-delta-variant-in-bd/BD_new_Covid_Cases.csv")
df.head()
```

```python
df.drop(['Unnamed: 3'], axis=1, inplace=True)
df.head()
```

```python
districts=df['Districts']
country ="Bangladesh"
```

# Extracting longitude and latitude for each district by Geopy

```python
lat=[]
long=[]
for district in districts:
    
    loc = geolocator.geocode(district +','+ country)
    if loc:
        lat.append(loc.latitude)
        long.append(loc.longitude)
    else:
        lat.append('loc.latitude')
        long.append('loc.longitude')
```

```python
df['Latitude']=lat
df['Longitude']=long
```

```python
df.head()
```

```python
f = folium.Figure(width=1100, height=1000)
m=folium.Map([23.8107, 90.4126], zoom_start=8).add_to(f)
```

### Red marker means cases over 5000, the Orange marker means cases between 5000 and 1500, and the Light-Red marker means cases under 1500

```python
import colorama
from colorama import Fore, Style
fg= folium.map.FeatureGroup()
for lat, long,district,case in zip(df.Latitude, df.Longitude,df.Districts,df.Total_Cases):    
#     text ='Total Confirmed case in '+ district+' is: '+str(case)
    text = folium.Html('Total Confirmed case in '+ '<b>'+district+'</b>'+' is:'+ '<b>'+str(case)+ '</b>', script=True)
    popup = folium.Popup(text, max_width=500)
    if case>=5000:
        folium.Marker(location=[lat, long], popup=popup,icon=folium.Icon(color='red',icon='info-sign')).add_to(m)
    elif 4999>case >=1500:
        folium.Marker(location=[lat, long], popup=popup,icon=folium.Icon(color='orange',icon='info-sign')).add_to(m)
    else:
        folium.Marker(location=[lat, long], popup=popup,icon=folium.Icon(color='lightred',icon='info-sign')).add_to(m)
m.add_child(fg)
```
