wine quality scaling

```python
def values(x):
    if x <= 5:
        x = 'low'
    elif x >5 and x <7:
        x = 'medium'
    else:
        x = 'high'
    
    return(x)
        
```

```python
df['level'] = df['quality'].apply(lambda x: values(x))
```

```python
df.head()
```

```python
#convert quality string into numerical variable and apply ordinal encoding method.
label = LabelEncoder()

quality_score  = label.fit_transform(df['level'])

print(quality_score)
print((label.classes_))
```

```python
# Removing Quality and level column
x = df.iloc[:,:-2]
x.head()
```

```python
# converting data into one format
standard = StandardScaler()

std_x = standard.fit_transform(x)
```

** 4. Split data** 

```python
x_train,x_test,y_train,y_test = train_test_split(std_x,quality_score,test_size = 0.20,random_state = 40)


print("Training data:{}".format(x_train.shape))
print("Test data:{}".format(x_test.shape))
```

```python
results = []
```

```python
## 5. implimenting model
```

```python
model.fit(x_train,y_train)
```

```python
prediction_x = model.predict(x_test)
```

**Random forest accuracy**

```python
accuracy = accuracy_score(prediction_x , y_test)
accuracy
```

```python
confusion_matrix(prediction_x, y_test)
```

```python
#  Build Linear Regression Analysis Model
learner = LinearRegression(); #initializing linear regression model

learner.fit(x_train,y_train); #training the linear regression model
y_predicted = learner.predict(x_test)
score=learner.score(x_test,y_test);#testing the linear regression model
```
