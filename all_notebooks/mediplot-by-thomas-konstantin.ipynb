          BCG Method of Administration
     </h2>

BCG is given as a single intradermal injection at the insertion of the deltoid.

Someone with a positive tuberculin reaction is not given BCG, because the risk of severe local inflammation and scarring is high.
    
The characteristic raised scar that BCG immunization leaves is often used as proof of prior immunization. This scar must be distinguished from that of smallpox vaccination, which it may resemble.
   https://en.wikipedia.org/wiki/BCG_vaccine#:~:text=BCG%20is%20given%20as%20a,the%20insertion%20of%20the%20deltoid.

The "BodyMap" class when created generates the silhouette of a human body.
The "generate" method must receive


1) two lists or NumPy array, the first list is a list of body parts (similar to listing the countries in choropleth plot), and a list of values corresponding to each body part.

2)cmap - a matplotlib colormap.

3) background - the background of the plot.
