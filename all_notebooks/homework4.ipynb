```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

dataset =  pd.read_csv("../input/energy/renewable-energy-stock-account-2007-18.csv")
dataset.head()
```

```python
dataset.columns
```

```python
#now I want to see how many missing values do we have 
missing_val_count_by_column = (dataset.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column > 0]) 
```

```python
#now I want to check which type of label can be for resource
dataset["resource"].unique()
```

```python
def label_fix_resource(value):
    if value == 'Biogas':
        return 0
    elif value == 'Coal':
        return 1
    elif value == 'Gas':
        return 2
    elif value == 'Geothermal':
        return 3
    elif value == 'Hydro':
        return 4
    elif value == 'Oil':
        return 5
    elif value == 'Renewable':
        return 6
    elif value == 'Solar':
        return 7
    elif value == 'Total generation':
        return 8
    elif value == 'Waste heat':
        return 9
    elif value == 'Wind':
        return 10
    elif value == 'Wood':
        return 11
    else: 
        return 12
```

```python
dataset['resource'] = dataset['resource'].apply(label_fix_resource)
dataset['resource'].unique()
dataset.head()
```

```python
#now I want to check which type of label can be for flag
dataset["flag"].unique()
```

```python
def label_fix_resource(value):
    if value == 'R':
        return 0
    elif value == 'F':
        return 1
    elif value == 'P':
        return 2
    else: 
        return 3
```

```python
dataset['flag'] = dataset['flag'].apply(label_fix_resource)
dataset['flag'].unique()
dataset.head()
```

```python
missing_val_count_by_column = (dataset.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column > 0]) 
```

```python
clean_data = dataset.drop(['variable', 'units', 'magnitude','source', 'data_value'], axis = 1)
# I don't want to deal with the large amount of columns, because this is only for understanding purposes

from sklearn.impute import SimpleImputer  # Imputer is for missing values

# Fill in the lines below: imputation
my_imputer = SimpleImputer()
datasetClean = pd.DataFrame(my_imputer.fit_transform(clean_data))
# Fill in the lines below: imputation removed column names; put them back
datasetClean.columns = clean_data.columns

missing_val_count_by_column = (datasetClean.isnull().sum())
print(missing_val_count_by_column[missing_val_count_by_column > 0]) 
# I have managed remove missing value rows 
```

```python
# let's train this thing 
X = clean_data.drop('year', axis=1)
y = clean_data['year']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)

from sklearn.svm import SVC  # support vector machine - linear
svclassifier = SVC(kernel='poly')
fitted = svclassifier.fit(X_train, y_train)

y_pred = svclassifier.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred)) 
```
