from https://github.com/block8437/gender.py/blob/master/gender.py

```python
import requests, json

def getGenders(names):
	url = ""
	cnt = 0
	if not isinstance(names,list):
		names = [names,]
	
	for name in names:
		if url == "":
			url = "name[0]=" + name
		else:
			cnt += 1
			url = url + "&name[" + str(cnt) + "]=" + name
		

	req = requests.get("https://api.genderize.io?" + url)
	results = json.loads(req.text)
	
	retrn = []
	for result in results:
		if result["gender"] is not None:
			retrn.append((result["gender"], result["probability"], result["count"]))
		else:
			retrn.append((u'None',u'0.0',0.0))
	return retrn

if __name__ == '__main__':
	print(getGenders(["Brian","Apple","Jessica","Zaeem","NotAName"]))
```

```python
names = original_dataset["Name"].tolist()
first_names = []
for name in range(len(names)):
    full_name = str(names[name])
    first_names.append(full_name.split()[0])
   
gender = []

for n in range(len(first_names)):
    gender.append(getGenders(first_names[n])[0][0])
```

```python
new_data = original_dataset

new_data = new_data.drop(["Rank", "Total Net Worth", "$ Last Change", "$ YTD Change", "Unnamed: 7", "Unnamed: 8", "Unnamed: 9", "Unnamed: 10"], axis=1)
new_data["Gender"] = gender
new_data = new_data[:498]

new_data.to_csv('500_richest_people_name_country_industry_gender.csv',index=False)
```
