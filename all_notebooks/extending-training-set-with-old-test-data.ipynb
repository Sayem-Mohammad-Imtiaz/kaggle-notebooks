Old test set has 62918 rows, more than the new test set. Let us compare the old test set with the new to see if there are any duplicates.

```python
def dataframe_similarity(df1, df2, which='both'):
    """Find rows which are the same between two DataFrames."""
    comparison_df = df1.merge(df2,
                              indicator=True,
                              how='outer')
    diff_df = comparison_df[comparison_df['_merge'] == which]
    return diff_df
```

```python
testsimilarities = dataframe_similarity(test, test_labelled)
testsimilarities
```
