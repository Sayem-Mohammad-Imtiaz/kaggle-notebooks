```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# read csv and analyse the first row of the data
data_customer = pd.read_csv('../input/mall-customers/Mall_Customers.csv')
data_customer.head()
```

```python
# change column name
data_customer = data_customer.rename(columns = {'Genre':'gender',
                                         'Age':'age', 
                                         'Annual Income (k$)':'annual_income',
                                         'Spending Score (1-100)':'spending_score'})
data_customer.head()
```

```python
# change the data type for gender from 'categoric' to 'numeric'
data_customer['gender'].replace(['Female', 'Male'], [0,1], inplace=True)
```

```python
# delete CustomerID and gender column because we didnt need it
x = data_customer.drop(['CustomerID', 'gender'], axis=1)

# check file to ensure that we get what we want
x.head()
```

```python
# make list of inertia
from sklearn.cluster import KMeans
clusters = []
for i in range (1, 11):
    km = KMeans(n_clusters = i).fit(x)
    clusters.append(km.inertia_)
clusters
```

```python
# plot inertia
fig,ax = plt.subplots(figsize = (8, 4))
sns.lineplot(x=list(range(1, 11)), y = clusters, ax=ax)
ax.set_title('Search Elbow')
ax.set_xlabel('Clusters')
ax.set_ylabel('Inertia')
```

```python
km5 = KMeans(n_clusters=5).fit(x)
```

```python
x['Labels']=km5.labels_
```

```python
plt.figure(figsize = (8,4))
sns.scatterplot(x['annual_income'], x['spending_score'], hue=x['Labels'], palette=sns.color_palette('hls',5))
plt.title('KMeans with 5 clusters')
plt
```
