*Meaningful tool improvement and understanding application: loc v.s. iloc*

**loc()**

```python
#loc() is a label based data selecting and includes the last element of range passed to it
display(df.loc[df.IF_VAT_EXCLUDED == 1])
#loc() as demonstrated above also allows for use of conditionals and boolean data
```

**iloc()**

```python
#iloc() is an indexed based selecting method that takes integer parameters to find row/col
print("Spontaneous iloc use to access a SALE_PRICE:") #look into wrapping outputs of a dataframe with a label
display(df.iloc[15,4])
print("\n==================================================================================")
#unlike loc() it doesn't accept boolean data within the paramaters
display(df.iloc[0:5, 0:5])
```

*Dropping any and all NaN values in the dataframe*

```python
df = df.dropna(axis=0)
#checks all columns in each row and then drops those rows
#as such axis=0 refers to checking and acting on rows within cols 
#conversely, axis=1 refers to checking and acting on the cols within the rows
df
```
