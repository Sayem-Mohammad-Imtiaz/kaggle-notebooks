## occurred at 14:46 JST (05:46 UTC) on March 11.

```python
import os
import numpy as np
import pandas as pd
import random
import folium
from folium import plugins
import seaborn as sns
import cv2
import matplotlib.pyplot as plt
import IPython
import time
import plotly.express as px
```

```python
data0=pd.read_csv('../input/earthquakes-in-japan/Japan earthquakes 2001 - 2018.csv')
data0[data0['mag']==9.1]
```

```python
print(data0.columns.tolist())
```

```python
data0=data0.fillna('N')
data1=data0[['time','latitude','longitude','depth','mag',]]
data1[0:5]
```

```python
# earthquakes seen in march, 2011
data2=pd.DataFrame(columns=data1.columns.tolist())
for i,item in enumerate(data1['time']):
    if item[0:4]=='2011'and item[5:7]=='03':
        data2=pd.concat([data2,data1.loc[i:i]])
data2=data2.sort_values(['time']).reset_index(drop=True)        
```

```python
from datetime import datetime as dt
data2a=data2.copy()
date=[]
for item in data2['time']:
    date+=[dt.strptime(item[0:-5],'%Y-%m-%dT%H:%M:%S')]
data2a['date']=date
data2a
```

## Is there any sign of 'the mag 9.1 earthquake' coming? 

```python
import plotly.express as px
fig=px.scatter(x=data2a['date'],y=data2a['mag'])
fig=px.scatter(data2a, x="date", y="mag")
fig.update_layout(title='Magnituide of earthquake in March, 2011')
fig.show()
```

### Focusing the period of March 9-11, the magnitude and frequency of earthquakes gradually decreased until the occurence of the event. Excessive energy  would be accumulated undersea for the event.

## The locations of earthuakes seen for 2 days before the event

```python
data3=data2a[['latitude','longitude',]]
start=dt.strptime('2011-03-09T00:00:00','%Y-%m-%dT%H:%M:%S')
end=dt.strptime('2011-03-11T05:47:00','%Y-%m-%dT%H:%M:%S')
```

```python
for i in range(len(data2a)):
    if start<data2a.loc[i,'date']<end:
        item=data2a.loc[i:i,'time'].tolist()[0][0:-5]
        eq_map = folium.Map(location=[37,137],tiles='Stamen Terrain',zoom_start=5.0,min_zoom=2.0)
        eq_map.add_child(plugins.HeatMap(data3.loc[i:i]))
        eq_map.save("./"+str(item)+".html")
```

```python
# earthquakes seen from 03-09 00:00:00 to 03-11 05:47:00
data4=data2a[data2a['date']<end][start<data2a['date']]
data4=data4.reset_index(drop=True)
data4
```

```python
data5=data4[['latitude','longitude',]]
eq_map = folium.Map(location=[37,137],tiles='Stamen Terrain',zoom_start=5.0,min_zoom=2.0)
eq_map.add_child(plugins.HeatMap(data5))
eq_map
```

###  Fifty five erathquakes were seen for 2 days before the event. Their locations were concentrated in one area, that was a neighbour of the place of 'the mag 9.1 earthquake'.

```python
paths0=[]
import os
for dirname, _, filenames in os.walk('./'):
    for filename in filenames:
        if filename[-5:]=='.html':
            paths0+=[os.path.join(dirname, filename)]
```

```python
order=[]
for item in paths0:
    order+=[item[:-5].split('/')[-1]]
paths1=pd.DataFrame(paths0)
paths1[1]=order
paths1.columns=['path','int']
paths1=paths1.sort_values('int')
paths2=paths1['path'].tolist()
```

```python
htmls=[]
for item in paths2:
    htmls+=["<iframe width='900' height='600' src='" +item+ "' frameborder='0'></iframe>"]
```
