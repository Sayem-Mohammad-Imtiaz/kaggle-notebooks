```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
filepath='../input/netflix-shows/netflix_titles.csv'
def import_dataset(path):
    df = pd.read_csv(path)
    return df
```

```python
df = import_dataset(filepath)
df.head()
```

```python
print(df.isna().sum())
```

```python
df['director'] = df['director'].fillna("")
```

```python
print(df.isna().sum())
```

```python
for i in df.columns:
    df[i] = df[i].fillna("")
    df[i] = df[i].astype(str)
```

```python
print(df.isna().sum())
```

```python
df.dtypes
```

```python
df = df[['title','director','cast','country','listed_in','description']]
df.head()
```

```python
for i in ['director','cast','country','description','listed_in']:
    df[i] = df[i].apply(lambda x:x.lower())
```

```python
df.head()
```

```python
df['director'] = df['director'].apply(lambda x: x.replace(','," "))
```

```python
df['cast'] = df['cast'].apply(lambda x: x.replace(','," "))
```

```python
df['listed_in'] = df['listed_in'].apply(lambda x: x.replace(','," "))
```

```python
df.head()
```

```python
df['total'] =df['title']+" "+df['listed_in']+" "+ df['director']+" "+df['cast']+" "+df['country']+" "+df['description']
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.tokenize import word_tokenize

tfv = CountVectorizer()
text_features = tfv.fit_transform(df['total'])
```

```python
similarity_matrix = cosine_similarity(text_features)
similarity_matrix
```

```python
import random
movie_title = random.sample(list(df['title']),1)
movie_title[0]
```

```python
def recommended_movie(movie_title):
    movie_index = df[df.title == movie_title].index
    similarity_score = similarity_matrix[movie_index]
    movie_frame = pd.DataFrame(similarity_score[0],columns=['cosine_similarity'])
    movie_frame=movie_frame.reset_index()
    movie_frame = movie_frame.sort_values(by='cosine_similarity',ascending=False)
    top10 = 10 
    top10_index = list(movie_frame['index'])[:10]
    recommended_movies = {}
    for i in top10_index:
        temp = df['title'][i]
        score = movie_frame['cosine_similarity'][i]
        recommended_movies[temp]=score
    print(recommended_movies)
```

```python
recommended_movie('The Seven Deadly Sins')
```
