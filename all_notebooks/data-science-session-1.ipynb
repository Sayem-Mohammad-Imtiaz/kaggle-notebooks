```python
## Data Structures
# List
# Tuples
# Dictionary
# Set

# List is a collection which is ordered and changeable. Allows duplicate members.
# Tuple is a collection which is ordered and unchangeable. Allows duplicate members.
# Set is a collection which is unordered and unindexed. No duplicate members.
# Dictionary is a collection which is unordered and changeable. No duplicate members.
```

```python
# List - Collections of elements
# Lists are used to store multiple items in a single variable.
# Lists are created using square brackets []

BFSI = ["IMA", "Axis", "HDFC"]
print(BFSI)
print(type(BFSI))
print(len(BFSI)) # len - no of elements in the list

DS_course = ['Data Science', 60, 'Python', True]
print(DS_course)
print(len(DS_course))

# List Items
# List items are ordered, changeable, and allow duplicate values.
# List items are indexed, the first item has index [0], the second item has index [1] etc.
# When we say that lists are ordered, it means that the items have a defined order, and that order will not change.
# If you add new items to a list, the new items will be placed at the end of the list.
```

```python
# Tuples : Operations are like list. Tuples are immutable (Tuples are unchangeable / immutable, meaning that we cannot change, add or remove items after the tuple has been created.)
# A tuple is a collection which is ordered and unchangeable and allow duplicate values.
# Tuples are written with round brackets.()
# Tuple items are indexed, the first item has index [0], the second item has index [1] etc.
# When we say that tuples are ordered, it means that the items have a defined order, and that order will not change.


BFSI = ("IMA", "Axis", "HDFC")
print(BFSI)
print(type(BFSI))
print(len(BFSI))
```

```python
# Set : To store unique values. does not store duplicates
# A set is a collection which is both unordered, unindexed,unchangeable and do not allow duplicate values.
# Sets are written with curly brackets.{}
# Unordered means that the items in a set do not have a defined order. Set items can appear in a different order every time you use them, and cannot be referred to by index or key.
# Once a set is created, you cannot change its items, but you can add new items.
# Sets cannot have two items with the same value.

BFSI = {"IMA", "Axis", "HDFC"}
print(BFSI)
print(type(BFSI))
print(len(BFSI)) 
```

```python
# Dictionary : Key & Value pairs
# Dictionaries are used to store data values in key:value pairs.
# A dictionary is a collection which is unordered, changeable and does not allow duplicates.
# Dictionaries are written with curly brackets, and have keys and values {BFSI:2}
# Dictionary items are unordered, changeable, and does not allow duplicates.
# Dictionary items are presented in key:value pairs, and can be referred to by using the key name.
# When we say that dictionaries are unordered, it means that the items does not have a defined order, you cannot refer to an item by using an index.
# Dictionaries are changeable, meaning that we can change, add or remove items after the dictionary has been created.
# Dictionaries cannot have two items with the same key

BFSI = {"IMA" : 4, "Axis": 3, "HDFC" : 2}
print(BFSI)
print(type(BFSI))
print(len(BFSI)) # len - no of elements in the list
```
