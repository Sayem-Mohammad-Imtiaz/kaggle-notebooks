#The impact of the BCG vaccination program in Hungary on the tuberculosis problem in children

Authors: I. Vadász & È. Demény  -Pneumonologie volume 151, pages151–159(1974)

Using the yearly standard data supply and data of some representative epidemiological investigations, the present authors studied the impact of the BCG vaccination program on tuberculosis in children during the last 20 years in Hungary.

Considering the incidence of active tuberculosis among vaccinated and unvaccinated children according to age as well as the rate of vaccination in the various age groups, authors point out that: 

The BCG vaccination program proved to be effective during the entire period of investigation;

Prevalence of bacillary TB had a marked influence on the incidence of the disease, both in vaccinated and unvaccinated children;

The impact of the BCG-vaccination program on the problem of childhood tuberculosis had been determined by the percentage of those vaccinated; 

And finally  the relatively high frequency of disease in the higher age groups (15 to 19-year-olds) indicates that the TB problem still exists and continuation of the various TB control methods is therefore  justified.https://link.springer.com/article/10.1007/BF02097162

```python
# cleaning df['hashtags']

# let us remove the non-Alphanumeric
df['text'] = df['text'].str.replace(r'\W',' ')# using Regural expression code to remove the +-=

text_list_clean = ['nan','nan, ',' ...,']
df['text'] = df['text'].str.replace('|'.join(text_list_clean),'')
```

```python
# Top 15 user location by number of tweets


import plotly.io as pio

pio.templates.default = "plotly_white"


data_plot = df.groupby('place_full_name')['screen_name'].count().reset_index()

data_plot = data_plot.sort_values(['screen_name'])

data_plot = data_plot.tail(15)

fig = px.bar(data_plot, x='place_full_name', y='screen_name',
             color='screen_name',labels={'screen_name':'Number of Screen Names'},
             title='Top 15 Screen Names places by number of tweets',
             height=600)
fig.show()
```

```python
from wordcloud import WordCloud, STOPWORDS

# let us plot hastag

comment_words = '' 
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in df['text']: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 1800, height = 1200, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = '#40e0d0') 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.title('text')  
plt.show()
```
