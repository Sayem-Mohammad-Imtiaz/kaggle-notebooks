```python
#all libraries
import tensorflow as tf
import os
import numpy as np # linear algebra
import pandas as pd 
from tqdm import tqdm
from matplotlib import pyplot as plt
import tensorflow as tf
# keras API for deep learning
from tensorflow import keras
# for image visulaizations
```

```python
# loading image path
image_path = []

roots, dirs, files in os.walk('../input/people-clothing-segmentation/png_images/IMAGES')
for file in files:
    path= os.path.join(roots, file)
    image_path.append(path)
    


mask_path = []

for roots_,b,c in os.walk('../input/people-clothing-segmentation/png_masks/MASKS'):
    for file in c:
        path= os.path.join(roots_, file)
        mask_path.append(path)


len(image_path),len(mask_path)
```

```python
image_path.sort()
mask_path.sort()
```

```python
# reading images
image=[]
for x in tqdm(image_path):
    path=tf.io.read_file(x)
    image.append(tf.image.decode_image(path, channels=3, dtype=tf.uint8))

    
    
mask=[]
for x in tqdm(mask_path):
    path=tf.io.read_file(x)
    mask.append(tf.image.decode_image(path, channels=1, dtype=tf.uint8))
len(image), len(mask)
```

```python
# Normalize Image
def normalize(input_image):
    input_image= tf.cast(input_image, tf.float32)/255.0
    return(input_image)

def load_image(inp,typ):
    input_image = tf.image.resize(inp, (128, 128))
    if typ=='image':
        input_image=normalize(input_image)
    elif typ=='mask':
        pass
    return input_image
```
