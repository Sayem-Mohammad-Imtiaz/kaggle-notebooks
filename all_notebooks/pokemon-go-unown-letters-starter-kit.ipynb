# You received a letter: "Keep Angry Give Give Laugh Engage!", Could you decode it and show for us?

```python
letter = "Keep Angry Give Give Laugh Engage!"
```

```python
exc = letter[-1]
```

```python
letter = letter[0:-1]
```

```python
letter
```

```python
exc
```

```python
words = letter.split(' ')
```

```python
decode = ''
for word in words:
    decode+=(df[df['pokemon_name'] == word]['letter'].values[0])
decode+=exc
```

```python
decode
```

## Let's visualize the decoded letter

```python
decode_imgs_path = []
for word in words:
    decode_imgs_path.append(df[df['pokemon_name'] == word]['img_name'].values[0])
decode_imgs_path.append(df[df['letter'] == exc]['img_name'].values[0])
```

```python
decode_imgs_path
```

```python
imgs = [Image.open(INPUT_PATH + img).resize((100,100), Image.ANTIALIAS)
        for img in decode_imgs_path]
```

```python
n_imgs = len(imgs)
```

```python
print(n_imgs)
```

```python
plt.figure(figsize=(10,10))

for i in range(0, n_imgs):

    plt.subplot(1, n_imgs, i+1) 
    plt.imshow(imgs[i])
    plt.axis('off')
    
plt.show()
```

# <p style="background-color:skyblue; font-family:newtimeroman; font-size:150%; text-align:center">7. Thanks!</p>

```python
letter = "Thanks!"
letter = letter.upper()
word = letter[0:-1]
exc = letter[-1]

decode_imgs_path = []
for each_letter in word:
    print(each_letter)
    decode_imgs_path.append(df[df['letter'] == each_letter]['img_name'].values[0])
decode_imgs_path.append(df[df['letter'] == exc]['img_name'].values[0])

imgs = [Image.open(INPUT_PATH + img).resize((100,100), Image.ANTIALIAS)
        for img in decode_imgs_path]

plt.figure(figsize=(10,10))

for i in range(0, n_imgs):

    plt.subplot(1, n_imgs, i+1) 
    plt.imshow(imgs[i])
    plt.axis('off')
    
plt.show()
```
