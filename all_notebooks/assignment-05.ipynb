```python
#reading the dataset
meta_df=pd.read_json("../input/amazon-reviews/meta_Appliances.json",lines=True)
meta_df
```

```python
#since the dataset includes millions of data, for this assignment, I am only working on the first 50 data. 
c_50=meta_df.iloc[:50]
c_50
#c_50.columns
```

```python
#since the price column includes a "$" symbol, I had to modify the cell with the item price value only, without the sign, to be able to plot it easily. 
c_50.price=c_50.price.str.replace("$","")
```

```python
#One of the price cells, also included text, so that needs to be modified too
def priceonly(x):
    try:
        x=float(x)
        return x
    except ValueError:
        return None
    
    
c_50.price=c_50.price.apply(lambda x: priceonly(x))
c_50
```

```python
#Interactive plot for first 50 items and their prices
import plotly.express as px
fig2= px.scatter(c_50, x=c_50['title'],y=c_50['price'],color=c_50['price'])
fig2.show()
```

```python
#Scatter plot for prices of the first 50 items. 
import matplotlib.pyplot as plt

x=list(range(50))
y=c_50[['price']]
plt.scatter(x,y, alpha=0.5,c='purple')
plt.xlabel('Serial Number')
plt.ylabel('Price')
```
