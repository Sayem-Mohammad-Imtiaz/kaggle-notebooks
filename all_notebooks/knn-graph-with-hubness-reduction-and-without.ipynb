```python
from skhubness.data import load_dexter
X , y = load_dexter ()
```

```python
X.shape, (X==0).sum() , (X==0).sum() / (X.shape[0]*X.shape[1])  
```

```python
#from sklearn.neighbors import NearestNeighbors
#nbrs = NearestNeighbors(n_neighbors=5).fit(X) #  algorithm='ball_tree' #  {‘mutual_proximity’, ‘local_scaling’, ‘dis_sim_local’, None}, 
from skhubness.neighbors import NearestNeighbors
nbrs = NearestNeighbors(n_neighbors=5,hubness = 'mutual_proximity' ).fit(X) #  algorithm='ball_tree' #  {‘mutual_proximity’, ‘local_scaling’, ‘dis_sim_local’, None}, 
distances, indices = nbrs.kneighbors(X)
indices
i = 1
edges = np.zeros( (0,2), dtype= int ) 
for i in range(1,5):
    ed = indices[:,[0,i]]
    edges = np.concatenate( (edges,ed), axis = 0 )

g = igraph.Graph( directed = True )
g.add_vertices(range(len(X)))
g.add_edges(edges )
    
l2 = (np.sign(y+1)).astype(int)
l3 = [int(t) for t in l2 ]
g.to_undirected(mode = 'collapse')
louvain_partition = g.community_multilevel()# weights=graph.es['weight'], return_levels=False)
modularity1 = g.modularity( l3 )#, weights=graph.es['weight'])
print("The modularity for partition is {}".format(modularity1))

louvain_partition = g.community_multilevel()# weights=graph.es['weight'], return_levels=False)
modularity1 = g.modularity( louvain_partition.membership)#, weights=graph.es['weight'])

print("The modularity for Louvain partition is {}".format(modularity1))

# For standard KNN 
# The modularity for partition is 0.26619381976429707
# The modularity for Louvain partition is 0.5151309642839581

# For 'mutual_proximity'
#The modularity for partition is 0.30429951595876115
#The modularity for Louvain partition is 0.5947172401787921
```
