# Feature Engineering

```python
def country(df):
  Player = df['Player']
  arr = Player.split('(')
  df['Player_name'] = arr[0].strip()

  arr1 = arr[1].split(')')

  if '/' in arr1[0]:
    arr2 = arr1[0].split('/')
    df['Country'] = arr2[1].strip()
  else:
    df['Country'] = arr1[0].strip()
  return df
```

```python
df = df.apply(country, axis=1)
```

```python
df.loc[df['Player_name'] == 'A Flintoff', 'Country'] = 'ENG'
df.loc[df['Player_name'] == 'SJ Harmison', 'Country'] = 'ENG'
```

```python
df.loc[df['Player_name'] == 'Mohammad Nawaz', 'Country'] = 'PAK'
```

```python
## dropping the Player attribute
df.drop('Player', axis=1, inplace=True)
```

```python
def notOut(df):
  if '*' in df['HS']:
    return 1
  return 0

df['max_not_out'] = df.apply(notOut, axis=1)
```

```python
columns = ['Inn', 'NO', 'Runs', 'HS', 'Avg', '100', '50', '0']

for col in columns:
  df[col] = df[col].str.replace('-', '0')

df['HS'] = df['HS'].str.replace('*','')
```

```python
columns = ['Inn', 'NO', 'Runs', 'HS',  '100', '50', '0']

for col in columns:
  df[col] = df[col].astype('int')
```

```python
df['Avg'] = df['Avg'].astype('float')
```

# EDA STARTS

```python
## country wise EDA
df1 = df.groupby('Country')['Runs'].sum()

plt.figure(figsize=(10,6))
plt.title('Most Runs Made by a Country')
plt.ylabel('Runs')
plt.xlabel('Country')
sns.barplot(x = df1.index, y = df1.values)
plt.show()
```

```python
## Number of 100s by a country


df2 = df.groupby('Country')['100'].sum()
plt.figure(figsize=(10,6))
plt.title('Most Centuries Made by a Country')
plt.ylabel('Centuries')
plt.xlabel('Country')
sns.barplot(x = df2.index, y = df2.values)
plt.show()
```

```python
## Number of Innings played vs Number of centuries scored

plt.figure(figsize=(8,6))
plt.xlabel('Number of Innings')
plt.xticks([i for i in range(25,400,25)])
plt.ylabel('Number of Centuries')
plt.yticks([i for i in range(10,55,10)])
sns.scatterplot(x = df.Inn, y = df['100'])
plt.show()
```

```python
## Top Players who has Top Average but also has scored more than 2000 runs

df1 = df[df.Runs > 2000]
df1 = df1.sort_values(by = 'Avg', ascending=False).head(10)

plt.figure(figsize=(14,6))
sns.barplot(x = df1.Player_name, y=df1.Avg)
plt.xlabel('Player Name')
plt.ylabel('Average')
plt.show()
```

```python
## centuries per Inning

df1 = df[df['Mat'] > 30]
df1['100/Inn'] = df1['100'] / df1['Inn']
df1['100/Inn'] = round(df1['100/Inn'],2)
df1 = df1.sort_values(by='100/Inn', ascending=False).head(10)

plt.figure(figsize=(14,6))
sns.barplot(x = df1.Player_name, y=df1['100/Inn'])
plt.xlabel('Player Name')
plt.ylabel('Centuries Scored per Inning')
plt.show()
```

```python
## Players who has scored most ducks in the cricket.

df1 = df.sort_values(by = '0', ascending=False).head(10)

plt.figure(figsize=(14,6))
sns.barplot(x = df1.Player_name, y=df1['0'])
plt.xlabel('Player Name')
plt.ylabel('Most Ducks scored by players')
plt.show()
```

```python
## Most 50s scored by players

df1 = df.sort_values(by = '50', ascending=False).head(10)

plt.figure(figsize=(16,6))
plt.title('Most 50s Scored by Players')
sns.barplot(x = df1.Player_name, y=df1['50'])
plt.xlabel('Player Name')
plt.ylabel('Number of 50s')
plt.show()
```

```python
plt.figure(figsize=(12,8))

plt.title('Highest Score vs Number of Innings')
sns.scatterplot(x=df.Inn, y=df.HS, hue=df.max_not_out)
plt.xlabel('Number of Innings')
plt.ylabel('Highest Score')
plt.show()
```
