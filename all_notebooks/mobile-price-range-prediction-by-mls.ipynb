```python
def bar_box_violin_plot(x, y, data, figsize=(16, 8)):
    plt.figure(figsize=figsize)
    ax1 = plt.subplot(131)
    ax1 = sns.barplot(x=x, y=y, data=data)
    ax1.set(title="Bar plot")
    ax2 = plt.subplot(132)
    ax2 = sns.boxplot(x=x, y=y, data=data)
    ax2.set(title="Box plot")
    ax3 = plt.subplot(133)
    ax3 = sns.violinplot(x=x, y=y, data=data)
    ax3.set(title="Violin plot")
    plt.show()
    
for i in range(20):
    bar_box_violin_plot(df.columns[20], df.columns[i], df)
```

# 3. Split Train Data

```python
from sklearn.model_selection import train_test_split
df = pd.read_csv("../input/mobile-price-classification/train.csv")
price = df.pop("price_range") # pull out only "price_range" columnm from df DataFrame, price is Series data
train = df[['battery_power', 'int_memory', 'mobile_wt', 'pc', 'px_height', 'px_width', 'ram', 'sc_w', 'wifi']]
(x_train, x_test, y_train, y_test) = train_test_split(train, price, test_size=0.2, random_state=0) # make train and test set from train.csv

# test data
df1 = pd.read_csv("../input/mobile-price-classification/test.csv")
test = df1[['battery_power', 'int_memory', 'mobile_wt', 'pc', 'px_height', 'px_width', 'ram', 'sc_w', 'wifi']]

# just check split sucessfully
print(x_train)
print("-"*50)
print(y_train)
```

```python
# just check split sucessfully
print(x_test)
print("-"*50)
print(y_test)
```

# 4. Create Model by various MLs

```python
# Final output is classify price_range, so use some classification models
from sklearn.metrics import confusion_matrix, accuracy_score
import xgboost as xgb

dtrain = xgb.DMatrix(x_train, label=y_train)
dtest = xgb.DMatrix(x_test, label=y_test)
xgb_param = {"eta":0.3, "max_depth":6, "objective":"multi:softmax", "num_class":4}
clf_xgb = xgb.train(xgb_param, dtrain)
y_pred = clf_xgb.predict(dtest)
conf_matrix = confusion_matrix(y_test, y_pred)
score = accuracy_score(y_test, y_pred)
print(f"XGBoost ACC score: {score:.3f}")
print("-"*60)
print("confusion matrix")
print(conf_matrix)
print("-"*60)
```

```python
import lightgbm as lgb

lgb_train = lgb.Dataset(x_train, y_train)
lgb_param = {"objective":"multiclass", "num_class":4}
clf_lgb = lgb.train(lgb_param, lgb_train)
y_pred = clf_lgb.predict(x_test)
y_pred_max = np.argmax(y_pred, axis=1) # very important!!!!! Need to investigate what exactly this means.

conf_matrix = confusion_matrix(y_test, y_pred_max)
score = accuracy_score(y_test, y_pred_max)
print(f"LightGBM ACC score: {score:.3f}")
print("-"*60)
print("confusion matrix")
print(conf_matrix)
print("-"*60)
```

```python
# import machine learning modules
from sklearn.svm import LinearSVC, SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from catboost import CatBoostClassifier

names = ["SVC", "KNeighborsClassifier", "SGDClassifier", "RandomForestClassifier", "CatBoostClassifier"]
models = [SVC(random_state=0), KNeighborsClassifier(), SGDClassifier(random_state=0), 
          RandomForestClassifier(random_state=0), CatBoostClassifier(logging_level='Silent', random_state=0)]

def MLs_visualization(name, model, x, y, y_pred):
    score = accuracy_score(y, y_pred)
    conf_matrix = confusion_matrix(y, y_pred)
    print("-"*60)
    print(f"{name} acc score: {score:.3f}")
    print("-"*60)
    print(conf_matrix) 
    # Pandas.Series method (y_test type is Pandas.Series)
    test_range = []
    for i in range(4):
        y_test_range = y_test.value_counts()[i]
        test_range.append(y_test_range)
    # numpy method (y_pred type is Numpy)
    pred_range = []
    for i in range(4):
        y_pred_range = np.count_nonzero(y_pred==i)
        pred_range.append(y_pred_range)
    # Show bar.plot
    left = np.arange(4)
    width = 0.3
    plt.bar(left-width/2, test_range, width, align="center", edgecolor="black", linewidth=1, label="actual test")
    for s, t in zip(left-width/2, test_range):
        plt.text(s, t, t, horizontalalignment="center", verticalalignment="bottom")
        #plt.text(s, t, t, ha="center", va="bottom") # this expresion is also OK.
    plt.bar(left+width/2, pred_range, width, align="center", edgecolor="black", linewidth=1, label="predicted")
    for s, t in zip(left+width/2, pred_range):
        plt.text(s, t, t, horizontalalignment="center", verticalalignment="bottom")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(np.arange(0, 4, 1))
    plt.title(f'{name} acc score: {score:.3f}')
    plt.show()
    # clear list
    test_range.clear()
    pred_range.clear()

for name, model in zip(names, models):
    model.fit(x_train, y_train)
    y_predict = model.predict(x_test)
    MLs_visualization(name, model, x_test, y_test, y_predict)
```
