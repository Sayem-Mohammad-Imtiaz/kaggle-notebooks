**Importing Data**

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../input/heart-attack-analysis-prediction-dataset/heart.csv")
```

### Histogram

```python
fig, axes= plt.subplots(7,2, figsize=(20,30) )
df.hist(ax=axes)
plt.show()
```

### Boxplot

```python
fig, axis= plt.subplots(figsize=(20,6) )
df.boxplot(ax=axis)
plt.show();
```

### Distplot

```python
sns.distplot(df['trtbps'],bins=100)
```

### Lineplot

```python
plt.figure(figsize=(16,8) )
sns.lineplot(data=df)
```

```python
sns.lineplot(data=df['age'],label='age')
sns.lineplot(data=df['chol'],label='chol')
sns.lineplot(data=df['trtbps'],label='trtbps')
sns.lineplot(data=df['thalachh'],label='thalachh');
# as many as data can take for representation in lineplot. 
```

### Barplot

```python
plt.figure(figsize=(10,8) )
sns.barplot(data=df)
```

### Heatmap

```python
plt.figure(figsize=(12,8) )
sns.heatmap(df.corr(), annot=True, cmap='coolwarm' )
```

### Scatter Plot, Regplot, Lmplot & Swarmplot

```python
plt.figure(figsize=(12,8) )
sns.scatterplot(data = df)
sns.regplot(x=df['trtbps'],y=df['chol'])
sns.lmplot(x='age',y='thalachh', data=df)
```

```python
sns.regplot(x=df['trtbps'],y=df['chol'])
```

```python
sns.lmplot(x='age',y='thalachh', data=df,)
```

```python
sns.swarmplot(x=df['fbs'],y=df['age']);
# The graph shows(fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
```
