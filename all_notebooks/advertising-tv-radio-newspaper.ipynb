```python
X = advertising[['TV','Radio','Newspaper']]
Y = advertising['Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size=0.7,random_state=100)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

# **Model Evaluation**

```python
model_coef = pd.DataFrame(model.coef_,X_test.columns,columns=['Coeffiecients'])
model_coef
```

Above result says if TV price increase by 1 unit it will affect the sales by **0.053** units

```python
Y_pred = model.predict(X_test)
```

Calculating Error

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(Y_test,Y_pred)
r_squared = r2_score(Y_test,Y_pred)
print('Mean squared', mse)
print('Predict square',r_squared)
```
