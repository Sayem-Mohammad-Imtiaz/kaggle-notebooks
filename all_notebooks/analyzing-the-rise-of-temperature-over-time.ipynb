```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Reading data.

```python
data = pd.read_csv("../input/daily-temperature-of-major-cities/city_temperature.csv").drop("State", axis=1).dropna()
data.head()
```

# Data selection
To see temperature increase we can use region, city, temperature and date. So country as parameter we don't need.

```python
data = data.drop("Country", axis=1)
data.head()
```

# Data modification
Now we have only data, which we need. Let's modify it how we need.

```python
regions = list( data["Region"].unique() )
cities = list( data["City"].unique() )
```

```python
for region in regions:
    data["Region"][ data["Region"]==region ] = regions.index(region)
for city in cities:
    data["City"][ data["City"]==city ] = cities.index(city)
data.head()
```

```python
data.astype("float")
```

# Solving tasks
**Task: how much is the temperature increase in different parts of the world over time?**

```python
regions = list(data["Region"].unique())
```

```python
dataList = []

for region in regions:
    dp = data["AvgTemperature"][ data["Region"]==region ]
    increases = list( np.array(dp) - np.array(dp.to_list()[1:]+[0]) )
    increases = ( (increases[::-1])[1:] )[::-1]
    dataPackage = pd.Series(increases, index=range(len(increases)))
    dataList.append(dataPackage)
```

```python
angles = []
for dl in dataList:
    tg = ( (dl*pd.Series(dl.index)).mean()-dl.mean()*pd.Series(dl.index).mean() )/( (dl*dl).mean()-pow(dl.mean(), 2) )
    angles.append(np.arctan(tg))
```

Part of the world with highest increase of temnerature.

```python
Regions = list( pd.read_csv("../input/daily-temperature-of-major-cities/city_temperature.csv")["Region"].unique() )
print( Regions[angles.index(min(angles))] )
```

Part of the world with lowest increase of temnerature.

```python
Regions = list( pd.read_csv("../input/daily-temperature-of-major-cities/city_temperature.csv")["Region"].unique() )
print( Regions[angles.index(max(angles))] )
```

**Task: which cities are seeing a rapid increase in temperature over time?**

```python
cities = list( data["City"].unique() )
```

```python
temperaturesPerYear = []
for city in cities:
    temperaturesPerYear.append( data[["Year", "AvgTemperature"]][ data["City"]==city ].groupby("Year").mean() )
```

```python
angles = []

for package in temperaturesPerYear:
    xy = pd.Series( np.array(package["AvgTemperature"]) * np.array(package.index) )
    x = pd.Series( np.array(package.index) )
    y = pd.Series( np.array(package["AvgTemperature"]) )
    x2 = pd.Series( np.array(package.index) * np.array(package.index) )
    
    angles.append(np.arctan( (xy.mean()-x.mean()*y.mean())/(x2.mean()-pow(x.mean(), 2)) ))

orderedAngles = angles
orderedAngles.sort(reverse=True)
```

TOP-5 of cities with highest mean temperature increase.

```python
indexes = []
for i in range(5):
    indexes.append( angles.index(orderedAngles[i]) )
```

```python
Cities = list( pd.read_csv("../input/daily-temperature-of-major-cities/city_temperature.csv")["City"].unique() )
for index in indexes:
    print(Cities[index])
```
