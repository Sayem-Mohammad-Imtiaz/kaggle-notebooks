```python
def get_digits(str1):
    try:
        c = ""
        for i in str1:
            if i.isdigit():
                c += i
        return c
    except:
        print(str1)
```

```python
def similarwords_f(str_segment):
    for city in engcity_names_lower:
        for asgncity_name in popularcitieslist[city]['eng']['assigned_original']:
            if fuzz.ratio(str_segment,asgncity_name)>85:
                if str_segment != asgncity_name:
                    popularcitieslist[city]['eng']['assigned_original_single_token_variant'].add(str_segment)
        for asgncity_name in popularcitieslist[city]['urdu']['assigned_original']:
            if fuzz.ratio(str_segment,asgncity_name)>85:
                if str_segment != asgncity_name:
                    popularcitieslist[city]['urdu']['assigned_original_single_token_variant'].add(str_segment)            
    return str_segment
```

```python
df_unique_city.city=df_unique_city.city.apply(lambda x: similarwords_f(x))
```

```python
def replace_with_popularcitieslist(str_segment):
    for city in engcity_names_lower:
        if str_segment == city:
            return city
        for city_name in popularcitieslist[city]['eng']['assigned_original']:
            if str_segment == city_name:
                return city
        for city_name in popularcitieslist[city]['eng']['available_original_as_it_is']:
            if str_segment == city_name:
                return city
        for city_name in popularcitieslist[city]['eng']['available_original_in_multi_token']:
            if str_segment == city_name:
                return city
        for city_name in popularcitieslist[city]['eng']['assigned_original_single_token_variant']:
            if str_segment == city_name:
                return city
        for city_name in popularcitieslist[city]['eng']['assigned_original_in_multi_token_variant']:
            if str_segment == city_name:
                return city
        for city_name in popularcitieslist[city]['urdu']['assigned_original']:
            if str_segment == city_name:
                return city
        for city_name in popularcitieslist[city]['urdu']['available_original_as_it_is']:
            if str_segment == city_name:
                return city
        for city_name in popularcitieslist[city]['urdu']['available_original_in_multi_token']:
            if str_segment == city_name:
                return city
        for city_name in popularcitieslist[city]['urdu']['assigned_original_single_token_variant']:
            if str_segment == city_name:
                return city
        for city_name in popularcitieslist[city]['urdu']['assigned_original_in_multi_token_variant']:
            if str_segment == city_name:
                return city
    return str_segment
```

```python
df_unique_city.city=df_unique_city.city.apply(lambda x: replace_with_popularcitieslist(x))
```
