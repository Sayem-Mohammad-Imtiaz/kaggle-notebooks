**Complex Data Structures in Python**

```python
# tuple - an immutable collection of items
tup = (1, 2, 3, 4)
for i in tup:
    print(i)
```

```python
# loop the tuple by index of the tuple & trying to change its values
for i in range(0, len(tup)):
    tup[i] += 10
    print(i, tup)
```

List - A complex data structure in Python

```python
# list - a mutable collection of items
lists = [1, 2, 3, 4, 5, 6]
for i in lists:
    print(i)
```

```python
# loop the list by index & trying to change its values
for i in range(0, len(lists)):
    lists[i] += 10
    print(i, lists)
```

Dictionary - A complex data structure in Python

```python
# dictionary: key:value store where keys are immutable, but values are mutable
d = {}
for i in range(10):
    d.update({i: "Value " + str(i)})
print(d)
```

```python
# functions can also return values & raise Exceptions (code raising alerts for incorrect input)
def operate_two_numbers(number1, number2, operation):
    op = operation.lower()
    if (op=="add"): return number1 + number2
    elif (op=="minus"): return number1 - number2
    elif (op=="multiply"): return number1 * number2
    elif (op=="divide"): return number1 / number2
    else: raise Exception("No such operation found!")
```

```python
operate_two_numbers(1, 2, "minus")
```

```python
operate_two_numbers(1, 2, "helo")
```

> **Data Science** - Linear Regression Model (Predicting House Price)

```python
# reading data from a CSV file
import pandas as pd

house_file_path = '../input/housesalesprediction/kc_house_data.csv'
dataset = pd.read_csv(house_file_path).dropna() # removing n/a values to clean the data

# show the first five rows along with headers
dataset.head()
```

**Checking for Correlation between the numerical variables**

```python
dataset.columns
```

```python
import seaborn as sns #Visualization
corr = dataset[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'floors', 'sqft_lot', 'yr_built', 'yr_renovated']].corr()
sns.heatmap(corr, cmap = 'Wistia', annot= True)
```

**Confirming the correlation between price & sqft_living**

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(20,7))
dataset.plot(x='price', y='sqft_living', kind='scatter', ax=ax)
plt.show()
```

**Data Science - Linear Regression Model Building**

```python
import numpy as np 

# space is independent variable and price is the dependent variable
space=dataset['sqft_living'] # deciding independent variable
price=dataset['price'] # deciding on dependent variable

x = np.array(space).reshape(-1, 1) # converting the shape of the data as per the required format of Linear Regression Model
y = np.array(price) # converting the shape of the data as per the required format of Linear Regression Model
```

```python
# Splitting the data into two subsets - Train data and Test data
# Training data will be used for training the model
from sklearn.model_selection import train_test_split 
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=1/3, random_state=0)
```

```python
# Training Linear Regression Model using Training Dataset
from sklearn.linear_model import LinearRegression 
regressor = LinearRegression()
regressor.fit(xtrain, ytrain)
```

```python
# Predicting the prices
pred = regressor.predict(xtest)
```

```python
# Visualizing the training Test Results 
plt.scatter(xtrain, ytrain, color= 'red')
plt.plot(xtrain, regressor.predict(xtrain), color = 'blue')
plt.title ("Visuals for Training Dataset")
plt.xlabel("Space")
plt.ylabel("Price")
plt.show()
```

```python
# Visualizing the Test Results 
plt.scatter(xtest, ytest, color= 'red')
plt.plot(xtrain, regressor.predict(xtrain), color = 'blue')
plt.title("Visuals for Test DataSet")
plt.xlabel("Space")
plt.ylabel("Price")
plt.show()
```

**Verifying Linear Regression Model Accuracy**

```python
from sklearn import metrics
print("MAE:", metrics.mean_absolute_error(ytest, pred))
print('MSE:', metrics.mean_squared_error(ytest, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(ytest, pred)))
```
