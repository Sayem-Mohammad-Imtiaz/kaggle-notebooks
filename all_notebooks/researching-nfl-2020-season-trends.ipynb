```python
nfl_2020_stats_df = nfl_2020_stats_df.drop(['PenaltyYards', 'PenaltyType', 'IsNoPlay', 'PenaltyTeam', 'IsPenaltyAccepted', 'IsTwoPointConversionSuccessful', 'IsTwoPointConversion', 'IsPenalty', 'IsFumble', 'IsMeasurement', 'IsChallengeReversed', 'IsChallenge', 'NextScore', 'SeriesFirstDown', 'ToGo', 'Second', 'Minute', 'Quarter', 'YardLine', ], axis=1)
nfl_2020_stats_df.head()
```

```python
nfl_2020_stats_df = nfl_2020_stats_df.dropna(subset=['OffenseTeam', 'DefenseTeam'])
nfl_2020_stats_df
```

```python
offensive_stats_team = nfl_2020_stats_df.sort_values(by=['GameDate', 'OffenseTeam'])
offensive_stats_team.head()
```

```python
offensive_stats_team = offensive_stats_team.set_index('OffenseTeam')
offensive_stats_team
```

```python
def select_offensive_team(df, offensive_team):
    """
    Select offensive team  bball_df3.loc['HOU']
    """
    offensive_team = offensive_team.upper()
    team = df.loc[offensive_team]
    return team

steelers = select_offensive_team(offensive_stats_team,'PIT')
```

```python
passes_to_JUJU = steelers.loc[:, ['GameDate','DefenseTeam', 'PlayType', 'Yards', 'Down', 'Formation', 'IsTouchdown', 'PassType','TeamWin', 'Description']][(steelers['Description'].str.contains('B.Roethlisberger'.upper())) &
(steelers['Description'].str.contains('J.Smith-Schuster'.upper()))]
passes_to_JUJU.head()
```

```python
def complete_incomplete_passes(quarterback_name,receiver_name):
    complete_incomplete_pass = steelers.loc[:, ['GameDate','DefenseTeam', 'PlayType', 'Yards', 'Down', 'Formation',\
                    'IsTouchdown','PassType','TeamWin', 'Description']][(steelers['Description']\
                    .str.contains(quarterback_name.upper())) & (steelers['Description']\
                    .str.contains(receiver_name.upper()))]
    complete_pass = steelers.loc[:, ['GameDate','DefenseTeam', 'PlayType', 'Yards', 'Down', 'Formation',\
                    'IsTouchdown','PassType','TeamWin', 'Description']][(steelers['Description']\
                    .str.contains(quarterback_name.upper())) & (steelers['Description']\
                    .str.contains(receiver_name.upper()))]
    return complete_incomplete_pass, receiver_name, quarterback_name

juju_stats_df,juju_name, qb_name  = complete_incomplete_passes('B.Roethlisberger', 'J.Smith-Schuster')
diJohnson_stats_df,diJohnson_name, qb_name  = complete_incomplete_passes('B.Roethlisberger', 'Di.Johnson')
ebron_stats_df,ebron_name, qb_name  = complete_incomplete_passes('B.Roethlisberger', 'E.Ebron')
claypool_stats_df,claypool_name, qb_name  = complete_incomplete_passes('B.Roethlisberger', 'C.Claypool')
```

```python
claypool_stats_df.head()
```

```python
def player_viz(player,name,downs):
    players_pass_distribution = player[downs].value_counts().sort_index(ascending=True)
    fig, axes = plt.subplots(figsize=(6, 4))
    axes.bar(players_pass_distribution.index, players_pass_distribution.values, align='center')   
    axes.set_ylabel('Complete/Incomplete')
    axes.set_title('Complete/Incomplete Targets')
    plt.show()

player_viz(claypool_stats_df, claypool_name, 'Down')    
  
```

```python
player_viz(ebron_stats_df, ebron_name, 'Down')  
```
