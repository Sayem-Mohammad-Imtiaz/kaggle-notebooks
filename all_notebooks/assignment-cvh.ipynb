# Customer Yearly Amount Spent  Model

```python
y = df['Yearly Amount Spent']
X = df[['Avg. Session Length', 'Time on App','Time on Website', 'Length of Membership']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.3 ,random_state=101)
```

# Linear Regression: y = f(X) 
* y = f(X) = coef1 X1 + coef2 X2 + ...

*(y = Yearly, X = Avg, TimeApp,TimeWeb, Length)*

```python
from sklearn.linear_model import LinearRegression 
```

```python
reg = LinearRegression().fit(X_train,y_train)
```

```python
coef=pd.DataFrame([X_train.columns, reg.coef_]).T
coef=coef.rename(columns={0:'Attributes',1:'Coefficients'})
coef
```

# Predicting Test Data

```python
ytest_pred = reg.predict(X_test)
```

```python
plt.scatter(y_test,ytest_pred)
plt.xlabel('Y Test')
plt.ylabel('Predicted Y')
```

# Evaluating the Model

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, ytest_pred))
print('MSE:', metrics.mean_squared_error(y_test, ytest_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, ytest_pred)))
print('R2 score : ',metrics.r2_score(y_test, ytest_pred))
print('Variance: ',metrics.explained_variance_score(y_test,ytest_pred))
```

```python
x_ax = range(len(X_test))
plt.plot(x_ax, y_test, lw=1, color="blue", label="original")
plt.plot(x_ax, ytest_pred, lw=0.8, color="red", label="predicted",marker="o", markersize=4)
plt.legend()
plt.show()
```

```python
from learntools.core import binder
binder.bind(globals())

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.model_selection import cross_val_score
from xgboost import XGBRegressor

# Set Matplotlib defaults
plt.style.use("seaborn-whitegrid")
plt.rc("figure", autolayout=True)
plt.rc(
    "axes",
    labelweight="bold",
    labelsize="large",
    titleweight="bold",
    titlesize=14,
    titlepad=10,
)


def score_dataset(X, y, model=XGBRegressor()):
    # Label encoding for categoricals
    for colname in X.select_dtypes(["category", "object"]):
        X[colname], _ = X[colname].factorize()
    # Metric for Housing competition is RMSLE (Root Mean Squared Log Error)
    score = cross_val_score(
        model, X, y, cv=5, scoring="neg_mean_squared_log_error",
    )
    score = -1 * score.mean()
    score = np.sqrt(score)
    return score


# Prepare data
df = pd.read_csv("../input/ecommerce-customers/Ecommerce Customers.csv")
```

```python
X = df.copy()
y = X.pop("Yearly Amount Spent")


# YOUR CODE HERE: Define a list of the features to be used for the clustering
features = [
    "Time on App",
    "Time on Website",
    "Avg. Session Length",
    "Length of Membership",
]

# Standardize
X_scaled = X.loc[:, features]
X_scaled = (X_scaled - X_scaled.mean(axis=0)) / X_scaled.std(axis=0)


# YOUR CODE HERE: Fit the KMeans model to X_scaled and create the cluster labels
kmeans = KMeans(n_clusters=10, n_init=10, random_state=0)
X["Cluster"] = kmeans.fit_predict(X_scaled)
```

```python
Xy = X.copy()
Xy["Cluster"] = Xy.Cluster.astype("category")
Xy["Yearly Amount Spent"] = y
sns.relplot(
    x="value", y="Yearly Amount Spent", hue="Cluster", col="variable",
    height=10, aspect=1, facet_kws={'sharex': False}, col_wrap=3,
    data=Xy.melt(
        value_vars=features, id_vars=["Yearly Amount Spent", "Cluster"],
    ),
);
```
