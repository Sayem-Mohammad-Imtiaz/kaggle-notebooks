# Data cleaning
**We have 18207 rows. by a above command we can see that 5 rows having missing data.**

```python
#df.isnull().sum()
missing_data=pd.DataFrame({'total_missing': fifa_data.isnull().sum(), 'perc_missing': (fifa_data.isnull().sum()/18207)*100})
missing_data
```

```python
#fifa_data.fillna(0,inplace=True)
fifa_data.dropna(inplace = True)
```

```python
fifa_data.info()
```

# Distribution of Age, Weight and Height.

```python
plt.figure(figsize=(15,20),facecolor='red',edgecolor='blue')
fig, axs = plt.subplots(1, 3, sharey=True, tight_layout=True)

# We can set the number of bins with the `bins` kwarg

axs[0].hist(fifa_data["Age"])

plt.xlabel="A"

axs[1].hist(fifa_data["Weight"])

axs[2].hist(fifa_data["Height"])

```

```python
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
fifa_data['Age'].value_counts()[:10].plot.bar(color=colors)
plt.title('Most Common Ages')
```

```python
from collections import Counter
print(Counter(fifa_data['Nationality'].values))
```

```python
#len(fifa_data["Nationality"].unique().tolist())
#len(fifa_data["Nationality"])
top_nat=fifa_data['Nationality'].value_counts()[:10]
top_nat
```
