```python
from sklearn.model_selection import train_test_split
# training data is for training
# testing data is for checking how we did
train_data, test_data = train_test_split(train, test_size=0.2)
```

```python
#Linear regression prediction

from sklearn.linear_model import LinearRegression

features = train_data[['LotArea',"YearBuilt",'1stFlrSF','FullBath',"BedroomAbvGr"]]

target = train_data[['SalePrice']]

test_features = np.array(test_data[['LotArea',"YearBuilt",'1stFlrSF','FullBath',"BedroomAbvGr"]])
test_targets = np.array(test_data['SalePrice'])

linear_regression = LinearRegression()

linear_regression.fit(features,target)

y_pred = linear_regression.predict(test_features)

y_pred

##change in cost per change in (fillin the blank)

#lot area = sqft of ground 94 cents 
#yearBuilt = for every year it up 770 dollars
# 1stsqrFT = 85 $/1stFLsqrft 
#full baths = all other things being equal(same same but different ) for every full bath price up 31k
# Bedroom = per bedroom 2700 $ therefore decreases that amount of sqft from the 1st floor sqft price

```

```python
errors = []
for i in range(len(y_pred)):
    errors.append(test_targets[i] - y_pred[i])
    
```

```python
from statistics import median 
abs_error = [abs(error) for error in errors]
print(sum(abs_error)/len(abs_error))
print(min(errors))
print(max(errors))
print(sum(errors)/len(errors))
print(median(abs_error))
print(linear_regression.coef_)
```

```python
import matplotlib.pyplot as plt

target = np.array(target).reshape((-1, 1))

model = LinearRegression(fit_intercept=False)

features = np.array(features[['LotArea']]).reshape((-1,1))

model.fit((features), target)
print(f"Score {model.score(features, target)}\nIntercept {model.intercept_}\nSlope {model.coef_}")

def abline(slope, intercept):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')

plt.scatter(features, target)
#lets take the feautures and plot it against our prediction,,,no multi 
plt.plot(features,model.predict(features))
#abline(model.coef_[0], model.intercept_)
plt.show()
```
