# Data Cleaning and  Preparation

##a. Correct the data. b. Replace, impute, or delete missing values


```python
#from previus code 
data.isnull().any()
```

```python
#data that have missinf broad impact values
print(data[data.broad_impact.isnull()])
```

```python
#creating new DataFrame to prevent erros on the original data due to manipulation
dataF=data.copy()
df=data.copy()
dataF.head()
```

```python
#filling NaN values on broad_impact with the mean of broad impact
dataF['broad_impact'] = dataF['broad_impact'].fillna(dataF['broad_impact'].mean())
df['broad_impact'] = df['broad_impact'].fillna(df['broad_impact'].mean())

print('Does some data have missing values?: \n'+str(dataF.isnull().any()))
```

```python
dataF.info()
```

```python
#cleaning data that may we only need
rank_universities=data[['country','institution','score','year']]
print(rank_universities.head(10))
```

## C. Apply feature scaling

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
encoder = LabelEncoder()
df['country'] = encoder.fit_transform(df['country'])
#removing unnecesary data
df.drop('institution', axis=1, inplace=True)
df.drop('year', axis=1, inplace=True)
scaler = StandardScaler()
df.head()
```

```python
#split the data set
scaler = StandardScaler()
y = df['world_rank']
X = df.drop('world_rank', axis=1)
X = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=101)
```
