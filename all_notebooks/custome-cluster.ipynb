```python
plt.scatter(data['Age'],data['Income'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```

```python
x=data[['Age','Income']]
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
ms=StandardScaler()
stand=ms.fit_transform(x)
```

```python
wt=[]
for i in range(1, 10):
    kmeans=KMeans(n_clusters=i,init= 'k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(stand)
    wt.append(kmeans.inertia_)
plt.plot(range(1, 10),wt)    
plt.title("Elbow method")
plt.xlabel("No of Clusters")
plt.ylabel("wt")
plt.show()
```

```python
kmeans=KMeans(n_clusters=5,random_state=0)
kmeans.fit(stand)
```

```python
prediction=kmeans.fit_predict(stand)
prediction
```

```python
new_data=data.copy()
new_data['cluster']=prediction
new_data.head()
```

```python
plt.scatter(new_data['Age'],new_data["Income"],c=new_data['cluster'],cmap='rainbow')
plt.xlabel('Age')
plt.ylabel('Income')
plt.show()
```
