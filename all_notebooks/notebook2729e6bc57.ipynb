```python
data = pd.read_csv('../input/random-salary-data-of-employes-age-wise/Salary_Data.csv')
data.head()
```

```python
X = data.iloc[:, 0].values.reshape(-1,1)
y = data.iloc[:, 1].values.reshape(-1,1)
```

```python
sns.heatmap(data.corr(), annot= True)
```

Test and Train Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
plt.scatter(X_train, y_train, color = 'orange')
plt.plot(X_train, regressor.predict(X_train), color = 'green')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(X_test, y_test, color = 'orange')
plt.plot(X_train, regressor.predict(X_train), color = 'green')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

Mean absolute error

```python
a=mean_absolute_error(y_test, y_pred)
print(a)
```

```python
b=mean_squared_error(y_test, y_pred)
print(b)
print(np.sqrt(b))
```

```python
from sklearn.metrics import r2_score
print("Mean absolute error: %.2f" % np.mean(np.absolute(y_pred - y_test)))
print("Mean square error: %.2f" %np.mean((y_pred - y_test))**2)
print("R2-score: %.2f" % r2_score(y_pred , y_test))
```

Visualizing Actual and Predicted values

```python
results = pd.DataFrame({
    'Year of experience': np.array(X_test).flatten(),
    'Actual': np.array(y_test).flatten(),
    'Predicted': np.array(y_pred.round(decimals = 2)).flatten(),
})

#results = pd.DataFrame(results)
results
```

```python
results= results.astype('float')
df1 = results.head(8)

df1.plot(kind='bar',figsize=(10,8))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```
