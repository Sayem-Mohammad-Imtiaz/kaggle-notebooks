## 8. Usain Earth <a class="anchor" id="Section8"></a>

Urusa 1.0.8

Let's do this!

```python
!rm spacetime.csv
from csv import writer
from csv import reader

recordtime = 'spacetime'

with open('workout.csv', 'r') as read_obj, \
        open('spacetime.csv', 'w', newline='') as write_obj:

    csv_reader = reader(read_obj)
    csv_writer = writer(write_obj)
    
    for row in csv_reader:
        
        row.append(recordtime)
        csv_writer.writerow(row)

pd.read_csv('spacetime.csv')
```

```python
extracolumn = ['23', '59', '59998', '945', '8/8/2021']

import csv
with open('spacetime.csv', 'r') as f:
    logcsv = [line for line in csv.reader(f)]

logcsv.append(extracolumn)

with open('spacetime.csv', 'w') as f:
    csv.writer(f).writerows(logcsv)

pd.read_csv('spacetime.csv')
```

```python
text = open("spacetime.csv", "r")
text = ''.join([i for i in text]) \
    .replace("spacetime", "dayone")
x = open("spacetime.csv","w")
x.writelines(text)
x.close()

pd.read_csv('spacetime.csv')
```
