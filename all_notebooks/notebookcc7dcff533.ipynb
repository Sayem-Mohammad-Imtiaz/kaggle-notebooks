```python
df = pd.read_csv('/kaggle/input/yellow-pages-pakistan/Yellow Pages Pakistan.csv')
df.head()
```

```python
df.drop_duplicates(subset = ['Business Name'], inplace = True) #Drop duplicates from Business Name Column
df.shape
```

```python
#find websites substring from Business description and append in Website column
```

```python
#Poular Categories and Business Type
plt.figure(figsize=(7,3))
sns.countplot(data = df, y = 'Category', order = df['Category'].value_counts().index[:15])
plt.title('Popular Categories')

#Poular Categories and Business Type
plt.figure(figsize=(13,7))
sns.countplot(data = df, y = 'Business Type', order = df['Business Type'].value_counts().index[:10])
plt.title('Popular Business Type')
```

```python
plt.figure(figsize=(5,2))
sns.countplot(data = df, y = 'City', order = df['City'].value_counts().index[:5])
plt.title('Top 5 City')

plt.figure(figsize=(5,2))
sns.countplot(data = df, y = 'Province', order = df['Province'].value_counts().index[:5])
plt.title('Top 5 Province')
```

```python
#count of empty values except Votes, Rating, Website Column
for col in df.columns:
    if col != 'Votes' and col != 'Rating'  and col !='Website':
        print(col, " : ", df[col].isna().sum())
```
