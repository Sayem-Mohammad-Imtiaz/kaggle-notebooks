```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statistics import *
```

**Splitting into train and test sets and keeping the test set aside**

```python
from sklearn.model_selection import train_test_split as tp
train,test=tp(df_dropped,test_size=0.2,stratify=df_dropped["Potability"])
```

```python
train.Potability.value_counts()/len(train)
```

```python
test.Potability.value_counts()/len(test)
```

**Just checking above whether the stratified split works well or not. I am learning along too**

```python
X_labels=["ph","Hardness","Solids","Chloramines","Sulfate","Conductivity","Organic_carbon","Trihalomethanes","Turbidity"]
y_label=["Potability"]
```

```python
x_data_train=pd.DataFrame(train[X_labels])
y_data_train=pd.DataFrame(train[y_label])
print(type(x_data_train))
print(type(y_data_train))
```

```python
#cross validation performance of models below
# SGDClassifier
from sklearn.linear_model import SGDClassifier as SG
from sklearn.model_selection import cross_val_score
model=SG(random_state=42)
cross_val_score(model,x_data_train,y_data_train,cv=10,scoring='accuracy')
```

```python
#Support Vector Machines
from sklearn import svm
model1=svm.SVC(kernel='linear')
cross_val_score(model1,x_data_train,y_data_train,cv=10,scoring='accuracy')
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression as lg
model2=lg()
cross_val_score(model2,x_data_train,y_data_train,cv=10,scoring='accuracy')
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier as dt
model3=dt(max_depth=3,random_state=42)
cross_val_score(model3,x_data_train,y_data_train,cv=10,scoring='accuracy')
```

**Fiting the data and performance of the data on test sets**

```python
x_data_test=pd.DataFrame(test[X_labels])
y_data_test=pd.DataFrame(test[y_label])
```

```python
#Decision Tree
model3.fit(x_data_train,y_data_train)
l=model3.predict(x_data_test)
check=y_data_test['Potability'].to_numpy()
correct=sum(l==check)
print(correct/len(check))
```

```python
#Logistic Regression
model2.fit(x_data_train,y_data_train)
l=model2.predict(x_data_test)
check=y_data_test['Potability'].to_numpy()
correct=sum(l==check)
print(correct/len(check))
```

```python
#Support Vector Machine
model1.fit(x_data_train,y_data_train)
l=model1.predict(x_data_test)
check=y_data_test['Potability'].to_numpy()
correct=sum(l==check)
print(correct/len(check))
```

```python
#Sgd Classifier
model.fit(x_data_train,y_data_train)
l=model.predict(x_data_test)
check=y_data_test['Potability'].to_numpy()
correct=sum(l==check)
print(correct/len(check))
```

**As we can see the decision tree model beats every other model with an accuracy of 63%**
