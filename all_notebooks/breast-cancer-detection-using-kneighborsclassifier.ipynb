```python
df = pd.read_csv('../input/breast-cancer-wisconsin-data/data.csv')
```

```python
df = df.drop('id', axis=1)
df = df.drop('Unnamed: 32', axis=1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.countplot(data=df, x=df["diagnosis"])
plt.title("Diagnosis")
plt.show()
```

```python
sns.pairplot(data=df)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['diagnosis'] = le.fit_transform(df['diagnosis'])
df['diagnosis'].head()
```

```python
x = df.drop(columns="diagnosis", axis=1).values
y = df['diagnosis'].values
```

```python
x
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
for i in range(1, 20):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train, y_train)
    pred = knn.predict(x_test)
    acc = accuracy_score(pred, y_test)
    print ('n_neighbour : {} -> Prediction : {}'.format(i, acc))
```

**KNeighborsClassifier gives the best result when the value of n_neighbors is 3**

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
pred = knn.predict(x_test)
acc = accuracy_score(pred, y_test)
print (acc)
```

```python
plt.plot(pred, '--r')
plt.plot(y_test, '--b')
plt.title('Comparison between predicted vs actual value')
plt.show()
```
