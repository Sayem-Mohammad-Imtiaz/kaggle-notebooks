```python
df=df.drop(["generation_hydro_pumped_storage_aggregated","forecast_wind_offshore_eday_ahead"], axis = 1)
df.stb.missing()
```

```python
df.dtypes
#Convert time data types to required data data types 
```

```python
#Set data type time to date data frame

df["time"]=pd.to_datetime(df["time"], utc=True)
df.dtypes
```

```python
df = df.set_index('time')
df
```

```python
#Set timeframe data for easier accessing
df["year"]=df.index.year
df["month"]=df.index.month
df["month_name"]=df.index.month_name()
df["weekdays"] = df.index.day_name()
```

Visualization of data 

```python
plt.style.use("ggplot")
cols_plot = ["generation_wind_onshore","generation_solar","generation_fossil_hard_coal","generation_fossil_gas"]
axes=df[cols_plot].loc["2018-01":"2018-12"].plot(marker=".", linestyle= "None", subplots= True, figsize=(11,9))
for ax in axes:
    ax.set_ylabel('Generation')
```

The figures above shows a quick visulaization of the data:
    
    Generation from solar tends to more during the summer, especially around July - Aug and less during the colder 
    seasons. 
    However, generation from wind energy shows less seasonal variation, with high generation around March 
    - April while electricity generation from gas shows no significant variations through out the year 

```python
fig, axes=plt.subplots(3, 1, figsize=(11,9), sharex=True)
for name, ax in zip(["generation_wind_onshore","generation_solar","generation_fossil_hard_coal","generation_fossil_gas"], axes):
    sns.boxplot(data=df, x='month_name', y=name, ax=ax )
    ax.set_ylabel('GWh')
    ax.set_xlabel("Months")
    ax.set_title(name)
```

```python
monthly_mean=df["price_actual"].resample("m").mean()
weekly_mean=df["price_actual"].resample("w").mean()
daily_mean=df["price_actual"].resample("d").mean()
```
