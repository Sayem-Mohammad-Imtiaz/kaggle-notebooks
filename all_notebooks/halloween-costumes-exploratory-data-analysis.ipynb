## No missing values

```python
ax = sns.countplot(dat['product name'])
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
ax.set_title('Different Prodcut Counts')
plt.show()
```

## Amounts of different products

```python
ax = sns.distplot(dat.rating,kde=False,bins=5,color='green')
ax.set_title('Rating Distirbution')
plt.show()
```

```python
wc = WordCloud(width=900,height=500,min_font_size=2).generate(' '.join(dat['product name'].str.lower().values))
plt.imshow(wc)
```

## Product Name Wordcloud

```python
wc = WordCloud(width=900,height=500,min_font_size=2).generate(' '.join(dat[dat['title'].notna()]['title'].str.lower().values))
plt.imshow(wc)
```

## Review Title Wordcloud

```python
titles = ' '.join(dat.Review_Title.str.lower().values)
titles = titles.split()
titles_freq = dict()
for title in titles:
    if title in titles_freq:
        titles_freq[title]+=1
    else:
        titles_freq[title]=1
        
        
titles_freq
```
