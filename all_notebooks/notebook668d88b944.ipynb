```python
# load data 
df = pd.read_csv('/kaggle/input/premier-league/results.csv')
len(df)
```

```python
df.head(5)
```

```python
# let's see the trend of results through the seasons.
# seems in last years away wins has passed over draws

namesgrouped=df.groupby(['season','result'])['result'].count()
namesgrouped
sperc=(namesgrouped *100/ namesgrouped.groupby(level=[0]).transform('sum'))
dfperc=sperc.to_frame()
dfperc.columns = ['perc']
dfperc.reset_index(inplace=True)
plt.figure(figsize=(12, 4))
plt.title("{} for each result".format('%'))
gr=sns.lineplot(x="season", y="perc", hue="result", data=dfperc)
gr.set_xticklabels(dfperc['season'].unique(), fontsize=10, rotation=65);
```

```python
# max goals scored in a match by a team as local
print ('max goals in a match by a team as local:',df['home_goals'].max())
df.loc[df['home_goals'].idxmax(),['season','home_team','home_goals','away_team','away_goals']]
```

```python
# max goals scored in a match by a team as visitor
print ('max goals in a match by a team as visitor:',df['away_goals'].max())
df.loc[df['away_goals'].idxmax(),['season','home_team','home_goals','away_team','away_goals']]
```

```python
df['total_goals'] = df['home_goals'] + df['away_goals']
# Number of total goals per season has increased over time
df.groupby('season')['total_goals'].sum().plot()
```

```python
# The average number of total goals per season is increasing
avg_goals_per_season = df.groupby('season')['total_goals'].mean().reset_index()
avg_goals_per_season['season'] = avg_goals_per_season['season'].map(lambda s: int(s[:4]))
sns.regplot(x='season', y='total_goals', data=avg_goals_per_season)
```
