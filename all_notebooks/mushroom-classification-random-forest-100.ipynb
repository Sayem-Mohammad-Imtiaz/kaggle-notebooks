```python
# Importing the libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold, GridSearchCV
```

```python
# Reading the dataset

data = pd.read_csv('../input/mushroom-classification/mushrooms.csv')
data.head()
```

```python
# Exploring the dataset

data.describe()
```

```python
# Check null values

data.isnull().sum(axis=1)
```

```python
# We can see that we do not have any missing value in this dataset
```

```python
# One hot encoding all the variables

k = data.keys()
for i in k:
    le = LabelEncoder()
    n = str(i) + "_n"
    data[n] = le.fit_transform(data[i])

for i in k:
    del data[i]
data.head()
```

```python
# with this method I have converted the values to numeric type
```

```python
# Defining dependant and independent variables

X = (data.drop("class_n", axis=1)).values
y = (data["class_n"]).values
```

```python
# spliting data in train/test sets and applying MLP
# Using cross validation method for spliting, and GridsearchCV to optimize the hyper-parameters.


kfold_f = StratifiedKFold(n_splits=2, shuffle=True, random_state=1)
score = []
for i, (train_index, test_index) in enumerate(kfold_f.split(X, y)):
    x_train, x_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    kfold = StratifiedKFold(n_splits=2, shuffle=True, random_state=1)
    model = RandomForestClassifier()
    param_grid = {
        "max_depth": [10, 20, 30, None],
        "n_estimators": [200, 400, 600],
    }

    gridsearch = GridSearchCV(
        estimator=model, param_grid=param_grid, scoring="accuracy", n_jobs=5, refit=True
    )

    gridsearch.fit(x_train, y_train)

    score.append(gridsearch.score(x_test, y_test))
    
np.mean(score, axis=0)*100
    
```
