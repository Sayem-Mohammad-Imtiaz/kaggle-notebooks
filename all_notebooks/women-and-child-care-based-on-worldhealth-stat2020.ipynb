```python
class WHS:
    def __init__(self,name,title):
        self.name=name
        self.title=title
 
    def working(self):
        self.name.rename(columns={'Dim1':'Sex','First Tooltip':'Life Expectancy'},inplace=True)
        self.name_Loc_groupby=self.name.groupby(by=['Location','Period'])['Life Expectancy'].mean()
        self.name_df=pd.DataFrame(self.name_Loc_groupby)
        self.name_df.rename(columns={'Life Expectancy':self.title},inplace=True)
        self.name_year_groupby=self.name.groupby(by=['Period'])['Life Expectancy'].mean()
        pd.DataFrame(self.name_year_groupby)
        pd.DataFrame(self.name_year_groupby).iplot(kind='bar',title=self.title)
        self.figure(self.name_df)

    def figure(self,name_df):
        sorted_values=self.name_df.index.get_level_values(1).sort_values()
        fig=px.choropleth(data_frame=self.name_df,
                          locations=self.name_df.index.get_level_values(0),
                          locationmode='country names',
                          color=self.name_df[self.title],
                          animation_frame=sorted_values,
                          animation_group=sorted_values)
        fig.update_layout(dict1={'title':self.title})
        fig.show() 
```

```python
adolescent=pd.read_csv('../input/who-worldhealth-statistics-2020-complete/adolescentBirthRate.csv')
whs=WHS(adolescent,'Adolescent_Birth_Rate')
whs.working()
```
