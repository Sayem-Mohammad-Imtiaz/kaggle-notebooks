```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import data
data = pd.read_csv("../input/biomechanical-features-of-orthopedic-patients/column_2C_weka.csv")
data.head()
```

- The class columns contain Abnormal and Normal. 
- I want to change this to 1 or 0.

```python
data["class"] = [1 if i=="Abnormal" else 0 for i in data["class"]]
data.head()
```

* I want to show you some data parts that are malign and benign

```python
malign = data[data["class"]==1]
benign = data[data["class"]==0]

plt.scatter(malign.pelvic_incidence, malign.degree_spondylolisthesis, color="red", label="malign")
plt.scatter(benign.pelvic_incidence, benign.degree_spondylolisthesis, color="green", label="benign")
plt.legend()
plt.xlabel("sacral_slope")
plt.ylabel("pelvic_radius")
plt.show()
```

* I create x and y data but x data not normalized so ı will normalize x data after ı select

```python
y = data["class"].values
x_data = data.drop(["class"],axis=1)
```

```python
# normalization
x = (x_data - np.min(x_data)) / (np.max(x_data) - np.min(x_data))
```

* and I separate the data into x and y

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=1)
```

* We completed the data processing

* Now we can create the KNN model :

* first of all import the model
* than create a model which is knn ( n_neighbours => this is K number)
* and the last step is fit the model with x_train and y_train

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train,y_train)
```

* I choose the number K to be 3
* and print the score

* but I don't know is this accuracy the best

```python
print("K=3 score : {}".format(knn.score(x_test,y_test)))
```

* we have to find best K value 
* we try some k values with for loop then choose the best one

```python
score_list = []
for k in range(1,25):
    knn2 = KNeighborsClassifier(n_neighbors = k)
    knn2.fit(x_train, y_train)
    score_list.append(knn2.score(x_test,y_test))
print("k = {} and max score : {}".format(score_list.index(max(score_list)), max(score_list)))
```

```python
#show the accuracy values with diffrent K values

plt.plot(range(1,25),score_list)
plt.xlabel("K values")
plt.ylabel("Accuracy")
plt.show()
```
