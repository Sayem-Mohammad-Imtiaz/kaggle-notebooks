# Standardising the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc.fit(df_Sugar_cnt.drop('Outcome',axis=1))
```

```python
features = sc.transform(df_Sugar_cnt.drop('Outcome',axis=1))
```

```python
feature_frame = pd.DataFrame(features,columns=df_Sugar_cnt.columns[:-1])
feature_frame
```

# Find the correlation between the Independent variables

```python

plt.figure(figsize=(10,10))
sns.heatmap(feature_frame.iloc[:,:].corr(),annot=True)
```

```python
sns.pairplot(df_Sugar_cnt, hue = 'Outcome')
```

# Split the training and test data

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(feature_frame,df_Sugar_cnt['Outcome'],
                                                    test_size=0.30)
```

# Call the KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 1)
```

# Train the KNN on the data and then perform testing on the sample

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

# Generate the confusion matrix and classfication report for measuring the accuracy

```python
from sklearn.metrics import classification_report , confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

# Calculate the error rate for different values of K

```python
error_rate = []

for i in range(1,50):
    knn   = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

# Plot the graph of error rate for all the values of K

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,50),error_rate,color='blue',linestyle='solid',marker='o',markerfacecolor='green',markersize=10)
plt.title('Error Rate v/s K value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
knn = KNeighborsClassifier(n_neighbors = 28)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

# Conclusion : k=28 is the best value of KNN
