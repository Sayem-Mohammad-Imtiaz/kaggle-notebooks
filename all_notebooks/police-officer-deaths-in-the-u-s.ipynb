# Machine Learning

## Encoding catigorical data to numeric
### Encoding columns state and cause and adding the encoded values as the new columns to DataFrame

```python
df['ccode'] = df.cause_short.astype('category').cat.codes
df['scode'] = df.state.astype('category').cat.codes
df.head()
```

# Seperating required columns

```python
mdf=df[['year','scode','ccode']]
mdf.head()
```

```python
year=mdf.drop('ccode',axis='columns')
year.head()
```

```python
cause=mdf.ccode
cause.head()
```

# K-nearest neighbors (KNN) classification
### Predictions using knn algorithim Since our targeted variable is categorical

```python
from sklearn import linear_model
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

## Training the Feature attributes year and state to predict the cause of death

```python
knn.fit(year, cause)
```

## Predicting cause of deaths by passing year and state(encoded code) 

```python
knn.predict([[2020,59]])
```

## Creating list of predictions

```python
mypre = [[1990,20], [2005,10],[2014,5],[2018,11]]
```

## Predictions of the array

```python
knn.predict(mypre)
```
