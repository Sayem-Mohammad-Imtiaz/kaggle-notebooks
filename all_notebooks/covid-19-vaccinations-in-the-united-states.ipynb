```python
us_vaccinations = pd.read_csv('/kaggle/input/usa-covid19-vaccinations/us_state_vaccinations.csv')
```

```python
us_vaccinations.info()
```

```python
us_vaccinations.head()
```

```python
us_vaccinations.isna().sum()
```

```python
us_vaccinations['date'] = pd.to_datetime(us_vaccinations['date'], format = '%Y-%m-%d')
```

```python
us_vaccinations.head()
```

```python
imp_null = SimpleImputer(missing_values=np.nan, strategy='constant',fill_value = 0)

us_vaccinations1 = pd.DataFrame(imp_null.fit_transform(us_vaccinations))
```

```python
us_vaccinations1.isna().sum()
```

```python
us_vaccinations1.columns = us_vaccinations.columns
us_vaccinations1
```

```python
len(us_vaccinations1['location'].unique())
```

```python
us_vaccinations1['location'].unique()
```

```python
# It is observed that 65 uniqeu value in location column but USA states quantitiy is lower than it
drop_element =  ['United States','Bureau of Prisons', 'Dept of Defense', 'Indian Health Svc',
             'Long Term Care', 'United States', 'Veterans Health']
us_vaccinations1.drop(us_vaccinations[us_vaccinations['location'].isin(drop_element)].index,
                     inplace=True)
us_vaccinations1['location'].unique()
```

```python
pivot_columns = us_vaccinations1[['date', 'location', 'total_distributed']]

pivot_table = pivot_columns.pivot_table(index ='date',columns = ('location'),aggfunc = sum)
```

```python
pivot_table = pivot_table.sort_values(by='date',ascending = False)
pivot_table = pivot_table.head(7)
pivot_table
```

```python
print(len(pivot_table.index))
column_count = len(pivot_table.columns)
```

```python
differences_last_first = []

for i in range(0,59):
    
    x =  pivot_table.iloc[0,i]
    y =  pivot_table.iloc[6,i]
    
    differences = x-y
    differences_last_first.append(differences)
    
    
            

```

```python
differences_last_first_columns = us_vaccinations1['location'].unique()
```

```python
distribution_count = list(map(list, zip(differences_last_first_columns,differences_last_first)))
```

```python
distribution_count_ = pd.DataFrame(distribution_count,columns = ('State','distribution_of_last_week'))
```

**Most Successful States as a Distribution Quantity**

```python
distribution_count = distribution_count_.sort_values('distribution_of_last_week', ascending = False)
distribution_count['date'] = pd.to_datetime('2021-04-08',format = '%Y-%m-%d')
distribution_count.head(10)
```

```python
plt.figure(figsize=(20,16))

distribution_count.head(5).plot(x = 'State',y = 'distribution_of_last_week',kind = 'barh')
```
