Import the data


```python
ufc = pd.read_csv("/kaggle/input/ufcdata/data.csv")
```

Ranking of first 20 UFC fighters with most Submission wins

```python
ufc.head(10)
```

```python
ufc = ufc[["R_fighter","B_fighter","Winner","R_win_by_Submission","B_win_by_Submission"]]
```

```python
red_fighter = pd.DataFrame(ufc.groupby(['R_fighter'],sort=False)["R_win_by_Submission"].max())
```

```python
blue_fighter = pd.DataFrame(ufc.groupby(["B_fighter"],sort=False)["B_win_by_Submission"].max())
```

```python
red_fighter['id'] = red_fighter.index
```

```python
red_fighter.count()
```

```python
red_fighter.set_index(np.arange(1514),inplace=True)
```

```python
red_fighter = red_fighter[["id","R_win_by_Submission"]]
```

```python
red_fighter['SubmissionW'] = red_fighter["R_win_by_Submission"]
```

```python
del red_fighter['R_win_by_Submission']
```

```python
blue_fighter['id'] = blue_fighter.index
```

```python
blue_fighter.count()
```

```python
blue_fighter.set_index(np.arange(1987),inplace=True)
```

```python
blue_fighter = blue_fighter[["id","B_win_by_Submission"]]
```

```python
blue_fighter["SubmissionW"] = blue_fighter["B_win_by_Submission"]
```

```python
MostSub = pd.concat([red_fighter,blue_fighter])
```

```python
MostSub = pd.DataFrame(MostSub.groupby(['id'],sort=False)['SubmissionW'].max())
```

```python
MostSub['id_fighter'] = MostSub.index
```

```python
MostSub.count()
```

```python
MostSub.set_index(np.arange(2139),inplace=True)
```

```python
MostSub.rename(columns={'id':'id_fighter'},inplace=True)
```

```python
MostSub = MostSub[['id_fighter','SubmissionW']]
```

```python
MostSub = MostSub.sort_values(by='SubmissionW',ascending=False)
```
