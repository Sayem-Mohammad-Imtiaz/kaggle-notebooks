```python
##One hot encoding

df_encode = pd.get_dummies(df)

df_encode.head()
```

#### Spliting

```python
#Taking weight Data
X = df_encode.drop('Weight', axis=1)
y = df_encode['Weight']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=100)
```

#### Modelling

```python
def metrics(y_true, y_pred):
    print('MAE:',mean_absolute_error(y_true, y_pred),'\n')
    print('MSE:', mean_squared_error(y_true, y_pred),'\n')
    print('RMSE:', np.sqrt(mean_squared_error(y_true, y_pred)),'\n')
    
def predictions(model, X_train=X_train, X_test=X_test, y_train=y_train, y_test=y_test):
    model.fit(X_train, y_train)
    train_pred = model.predict(X_train)
    test_pred = model.predict(X_test)
    actual = [ y_train, y_test]
    pred = [train_pred, test_pred]
    data = ['Train', 'Test']
    
    for i in range(2):
        print(data[i])
        metrics(actual[i], pred[i])
    sns.scatterplot(x=y_test, y=test_pred)
    plt.show()
```

#### Linear Regression

```python
lr = LinearRegression()

test_pred=predictions(lr)
```

#### KNN

```python
knn = KNeighborsRegressor()

predictions(knn)
```

#### Decision Tree

```python
dtree = DecisionTreeRegressor()

predictions(dtree)
```

#### Random Forest

```python
rf = RandomForestRegressor()

predictions(rf)
```

#### Gradient Boosting

```python
grad = GradientBoostingRegressor()

predictions(grad)
```

#### 

#### ADA Boost

```python
ada = AdaBoostRegressor()

predictions(ada)
```

#### Light GBM

```python
li = LGBMRegressor()
predictions(li)
```

#### XGBoost

```python
xgb = XGBRegressor()

predictions(xgb)
```

#### CAT Boost

```python
cat = CatBoostRegressor()

predictions(cat)
```
