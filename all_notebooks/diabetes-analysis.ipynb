```python
# Import our visualization libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
color = sns.color_palette()
get_ipython().run_line_magic('matplotlib', 'inline')
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
import plotly.express as px
```

Let's try to explore the data with some visual representations. 
First, we have a look at what percentage have diabetes.

```python
dist = df['Outcome'].value_counts()
colors = ['lightblue', 'darkorange']
trace = go.Pie(values=(np.array(dist)),labels=dist.index)
layout = go.Layout(title='Diabetes Outcome')
data = [trace]
fig = go.Figure(trace,layout)
fig.update_traces(marker=dict(colors=colors, line=dict(color='#000000', width=3)))
fig.show()
```

Let's use a heatmap to see if any correlations possibly exist.

```python
def df_to_plotly(df):
    return {'z': df.values.tolist(),
            'x': df.columns.tolist(),
            'y': df.index.tolist() }
import plotly.graph_objects as go
dfNew = df.corr()
fig = go.Figure(data=go.Heatmap(df_to_plotly(dfNew)))
fig.show()
```

Here we can see that Glucose-Outcome and SkinThickness-Insulin and BMI-SkinThickness (to a lesser degree exhibit some correlation.

The next thing we want to look at is something we believe to have a correlative relationship. Let's see how Insulin-Glucose are represented visually. 

```python
fig = px.scatter(df, x='Glucose', y='Insulin')
fig.update_traces(marker_color="hotpink",marker_line_color='rgb(7,38,107)',
                  marker_line_width=2.5)
fig.update_layout(title_text='Glucose - Insulin')

fig.show()
```

We see, as expected, a relationship between glucose and insulin. We can also note common clustering between 75-130 in glucose, which maps to 0-200 for insulin levels. Are these representative of the non-diabetics in the population?

Next, let's have a look at how age plays into the equation. Let's try to use a boxplot this time. 

```python
fig = px.box(df, x='Outcome', y='Age')
fig.update_traces(marker_color="goldenrod",marker_line_color='rgb(6,58,107)',
                  marker_line_width=2.5)
fig.update_layout(title_text='Age - Outcome')
fig.show()
```

Ok, so it looks like the average age range for those with diabetes is higher than for non-diabetic.
Of course we expected some outliers here and we see numbers of non-diabetics in the 60-70 age group. 

```python
plot = sns.boxplot(x='Outcome',y="BMI",data=df)
```
