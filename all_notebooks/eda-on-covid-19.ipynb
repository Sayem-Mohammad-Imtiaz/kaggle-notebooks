```python
df = pd.read_csv("../input/latest-covid19-india-statewise-data/Latest Covid-19 India Status.csv")
df.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
total_states = np.arange(len(df['State/UTs']))
```

```python
total_states
```

# Total Active Cases

```python
from matplotlib.pyplot import figure
figure(num=None,figsize=(8,10), facecolor='w', edgecolor = 'k')
plt.barh(total_states,df.Active, align='center', alpha=0.5,color=(1,0,0), edgecolor=(0.5,0.2,0.8))

plt.yticks(total_states, df['State/UTs'])
plt.xlim(1,max(df.Active) +10)
plt.xlabel('Active Number of Cases')
plt.title('Corona Virus Cases')
plt.show()
```

# Total Deaths 

```python
from matplotlib.pyplot import figure
figure(num=None,figsize=(8,10), facecolor='w', edgecolor = 'k')
plt.barh(total_states,df.Deaths, align='center', alpha=0.5,color=(1,0,0), edgecolor=(0.5,0.2,0.8))

plt.yticks(total_states, df['State/UTs'])
plt.xlim(1,max(df.Deaths) +10)
plt.xlabel('Deaths Number of Cases')
plt.title('Corona Virus Cases')
plt.show()
```
