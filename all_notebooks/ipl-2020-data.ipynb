# Runs vs ball in over

Nothing interesting here, except the fact that some overs had 9 balls in an over. Just wanted to use scatter plot with plotly.

```python
balls = data['Ball'].apply(lambda n: n+(np.random.random_sample()-0.5))
runs = data['Total_Runs'].apply(lambda n: n+(np.random.random_sample()-0.5))
```

```python
sns.scatterplot(x= balls, y=runs,y_jitter=50, x_jitter=50)
plt.show()
```

```python
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
data = [go.Scatter(x= balls, 
                   y=runs,
                   mode='markers',
                   marker=dict(
                                size=2,
                                color='rgb(51,204,153)',
                               symbol = 'circle',
                               line=dict(width=2)
                            ),
                   opacity=0.9,
                   
                )]

layout = go.Layout(title='Runs vs ball in over',
                  xaxis = dict(title='Ball of the over'),
                  yaxis = dict(title='Runs'),
                  hovermode = 'closest')

fig = go.Figure(data=data, layout=layout)

pyo.iplot(fig)
```
