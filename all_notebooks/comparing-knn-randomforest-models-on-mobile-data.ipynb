# Importing Dataset

```python
train = pd.read_csv('../input/mobile-price-classification/train.csv')
test = pd.read_csv('../input/mobile-price-classification/test.csv')
```

# Dividing Training data into x and y

```python
x_train = train.iloc[:,0:-1]
x_train.shape
```

```python
y_train = train.iloc[:,-1]
y_train.shape
```

# Dividing Testing data into x and y

```python
x_test = test.iloc[:,0:-1]
x_test.shape
```

```python
y_test = test.iloc[:,-1]
y_test.shape
```

# Checking accuracy for Training data

```python
from sklearn.model_selection import train_test_split

x_train_1, x_test_1, y_train_1, y_test_1 =  train_test_split(x_train,y_train , random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold

knn = KNeighborsClassifier(n_neighbors = 5)

knn.fit(x_train_1, y_train_1)

kfold = KFold(n_splits = 10)

scores_1 = cross_val_score(knn, x_test_1, y_test_1, cv=kfold, scoring='accuracy');scores_1
```

```python
scores_1.mean()
```

# Conclusion: Training data is giving 89% accuracy

# Using KNN model with CV to check Accuracy on Test Data (Model 1)

```python
knn.fit(x_train, y_train)
```

```python
kfold = KFold(n_splits = 10)

scores = cross_val_score(knn, x_test, y_test, cv=kfold, scoring='accuracy');scores
```

```python
scores.mean()
```

# Conclusion : We can see that our model is giving 49% accuracy for testing data which makes it a bad model.

# Using RandomForest model with CV to check Accuracy on Test data(Model 2)

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

n_esti = list(range(5,16))
max_dep = list(range(5,16))

rf=RandomForestClassifier()
rf.fit(x_train,y_train)
rf_param_grid = {'n_estimators' : n_esti, 'max_depth' : max_dep}
rf_grid = GridSearchCV(estimator=rf,
                      param_grid=rf_param_grid,
                      scoring='accuracy',
                      cv=4,
                      return_train_score=True)

rf_grid.fit(x_test, y_test)
```

```python
rf_grid_df = pd.DataFrame(rf_grid.cv_results_)
rf_grid_df.loc[rf_grid_df['rank_test_score'] == 1, ]
```

# Conclusion: We can see that our this model is also giving 53% accuracy for testing data which makes it a bad model.

# Based on our predictions we can see that on training data we are getting 89% accuracy while on testing data we are getting 49% and 53% as accuracy rates from two different models. This states that our model is not working great on testing data. Hence, it is a bad model.
