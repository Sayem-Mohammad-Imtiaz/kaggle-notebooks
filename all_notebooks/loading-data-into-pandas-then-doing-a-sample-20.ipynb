```python
import csv
tsv_filename = "/kaggle/input/clash-royale-season-18-dec-0320-dataset/BattlesStaging_12272020_WL_tagged/battlesStaging_12272020_WL_tagged.csv"
chunk_size = 500000

text_file_reader = pd.read_csv(tsv_filename, engine='python',encoding='utf-8-sig', quoting=csv.QUOTE_MINIMAL, chunksize = chunk_size)
```

```python
dfList = []
counter = 0

for df in text_file_reader:
    dfList.append(df)
    counter= counter +1
    print("Max rows read: " + str(chunk_size * counter) )

df = pd.concat(dfList,sort=False)
```

### Let's see the distribution of the average starting trophies of the players in a match

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter

ax = df.hist(column='average.startingTrophies', bins=25, grid=False, figsize=(12,8), color='#86bf91', zorder=2, rwidth=0.9)

ax = ax[0]
for x in ax:

    # Despine
    x.spines['right'].set_visible(False)
    x.spines['top'].set_visible(False)
    x.spines['left'].set_visible(False)

    # Switch off ticks
    x.tick_params(axis="both", which="both", bottom="off", top="off", labelbottom="on", left="off", right="off", labelleft="on")

    # Draw horizontal axis lines
    vals = x.get_yticks()
    for tick in vals:
        x.axhline(y=tick, linestyle='dashed', alpha=0.4, color='#eeeeee', zorder=1)

    # Remove title
    x.set_title("")

    # Set x-axis label
    x.set_xlabel("Average Trophies Distribution", labelpad=20, weight='bold', size=12)

    # Set y-axis label
    x.set_ylabel("Matches", labelpad=20, weight='bold', size=12)

    # Format y-axis label
    x.yaxis.set_major_formatter(StrMethodFormatter('{x:,g}'))
```
