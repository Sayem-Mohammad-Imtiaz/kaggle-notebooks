```python
import pandas as pd

main_file_path = '../input/house-prices-advanced-regression-techniques/train.csv' # this is the path to the Iowa data that you will use
data = pd.read_csv(main_file_path)

# Run this code block with the control-enter keys on your keyboard. Or click the blue botton on the left
print(data.describe())
```

```python
print(data.columns)
```

```python
print(data.head())
```

```python
intreast_data=data[['Id','SalePrice']]
intreast_data.describe()
```

```python
y = data.SalePrice
```

```python
lowa_pred= ['MSSubClass', 'YrSold', 'LotArea', 'YearRemodAdd', 'BsmtFinSF1']
X=data[lowa_pred]
```

```python
from sklearn.tree import DecisionTreeRegressor
lowa_model = DecisionTreeRegressor()
lowa_model.fit(X, y)
```

```python
print("Making predictions for the following 5 houses:")
print(X.head(11))
print("The predictions are")
print(lowa_model.predict(X.head(11)))
```
