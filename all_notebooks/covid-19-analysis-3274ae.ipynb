## remove world from dataframe

```python
data = data[ (data['location'] != 'World') & (data['location'] != 'International')]
data
```

## exclude negative new cases

## you can see above that 17 rows have negative new cases, which is incorrect because it is not possible to have negative daily cases. The minimum should be zero

```python
data = data[ data['new_cases'] >= 0 ]
```

```python
data['new_cases'].min()
```

```python
summry_df = data.groupby(['location'])['new_cases'].sum().reset_index().sort_values(['location'])
summry_df
```

```python
summry_df[summry_df ['location'] == 'United Arab Emirates']
```

```python
summry_df2 = summry_df = data.groupby(['location'])['new_cases', 'new_tests'].sum().reset_index().sort_values(['location'])
summry_df2
```

```python
summry_df2[summry_df2 ['location'] == 'United Arab Emirates']
```

```python
data[ data['location'] == 'United Arab Emirates']['population'].reset_index()
```

```python
summry_df3 = summry_df = data.groupby(['location'])['new_cases', 'new_tests', 'population'].sum().reset_index().sort_values(['location'])
summry_df3
```

## For population we can not sum we need the average

```python
summry_df4 = summry_df = data.groupby(['location']).agg(
    {
        'new_cases': np.sum,
        'new_tests': np.sum,
        'population': np.mean
    })
summry_df4.reset_index()
```
