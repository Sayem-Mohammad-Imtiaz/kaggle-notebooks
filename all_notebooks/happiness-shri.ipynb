The World happiness survey aims to study happiness and factors contributing to it across the globe. It's interesting to note that the report aims to quanitify a subjective term like happiness. For convenience, I have only considered the data for 2019. 

```python
#Importing required libraries
import numpy as np
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset=pd.read_csv('/kaggle/input/world-happiness/2019.csv')
```

There are six factors on the basis of which all countries are ranked:
1. GDP per capita
2. Healthy life expectancy
3. Freedom to make life choices
4. Generosity
5. Perceptions of corruption

```python
dataset.describe()
```

```python
#Checking for missing data
dataset.isnull().sum()
```

```python
dataset.head()
```

For the purpose of understanding relationship between Happiness score with different variables,let's use different visualization techniques.
First, we shall depict using heatmaps

```python
#correlation matrix
corrmat = dataset.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.10, square=True,annot=True);
```

From the visualization given above, we observe that **GDP per capita**,**Social support** and **Healthy life expectancy** are strongly correlated with the Happiness score. Freedom to make life choices also relates well with happiness score to a certain extent. Let's analyze them one by one

```python
#Observing the distribution of the happiness score data
sns.distplot(dataset['Score'])
```

```python
# Scatterplots to understand relationship of Score with the selected 3 predictors 
sns.set()
cols = ['Score', 'GDP per capita', 'Social support', 'Healthy life expectancy']
sns.pairplot(dataset[cols])#, size = 2.5)
```

We observe in the above scatterplots that all 3 factors are positively correlated with the happiness score which means an increase in these factors contributes to an increase in the happiness score.Let us now pick the top 10 countries with a high happiness score and observe the influence of each of these characteristics on it.

```python
#Sorting the data and picking the top 10 countries with highest happiness score
sorted_df=dataset.sort_values(by='Score', ascending=False).head(10)
sorted_df
```

We have to reshape our existing dataframe inorder to observe the impact of each predictor across countries. this will give us an idea of which factor provides maximum contribution to overall score across countries

```python
top10=sorted_df[['Country or region','GDP per capita','Social support','Healthy life expectancy','Score']]
top10=top10.melt(id_vars=['Country or region','Score'],var_name='Parameter' , ignore_index=False)
```

```python
f=sns.catplot(x='Country or region', y='value', hue='Parameter', data=top10, kind='bar')
f.fig.set_figwidth(20)
f.fig.set_figheight(10)
```

Based on the plot above for the Top 10 countries with highest over all score we can infer that a high score of Social Support variable is a key contributor to a high happiness score.
