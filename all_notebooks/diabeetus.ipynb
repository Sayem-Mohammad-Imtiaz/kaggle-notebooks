# Split test v train

```python
# Splitting feature & target cols

df_X = df[[col for col in df.columns if 'Outcome' not in col]]
df_Y = df['Outcome']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_X, df_Y, test_size=0.25, random_state=101)
```

```python
# Recombining and cleaning data for fitting

df_train = X_train

df_train['Outcome'] = y_train

df_train = df_train[df_train['Pregnancies'] <= 12.5]
df_train = df_train[df_train['Glucose'] >= 50]
df_train = df_train[df_train['SkinThickness'] <= 60]
df_train = df_train[df_train['Insulin'] <= 350]
df_train = df_train[(df_train['BMI'] <= 50)&(df_train['BMI'] >= 18)]
df_train = df_train[df_train['DiabetesPedigreeFunction'] <= 1.2]
df_train = df_train[df_train['Age'] <= 65]

X_train = df_train[[col for col in df_train.columns if 'Outcome' not in col]]
y_train = df_train['Outcome']
```

```python
# from sklearn.preprocessing import StandardScaler
# sc=StandardScaler()
# X_train = sc.fit_transform(X_train)
# X_test = sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression as LogReg

logregmodel = LogReg()

logregmodel.fit(X_train, y_train)

predictions_logreg = logregmodel.predict(X_test)
```

# Model Validation

```python
import sklearn.metrics

sklearn.metrics.accuracy_score(y_test, predictions_logreg)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,predictions_logreg))
```
