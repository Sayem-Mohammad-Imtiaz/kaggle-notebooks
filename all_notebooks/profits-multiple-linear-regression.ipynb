# Profits Multiple Linear Regression

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

from sklearn.metrics import r2_score

import warnings
warnings.filterwarnings('ignore')
```

# Dataset

```python
df = pd.read_csv('/kaggle/input/50-startups/50_Startups.csv')
df.head()

X = df.iloc[:, :-1]    # independent varible
y = df.iloc[:, -1]     # dependent variable

X.head()

df['State'].unique()

# Since we do not need all the 3 columns and can work with only 2 columns for classifying the states, we will drop one column
states = pd.get_dummies(X['State'], drop_first=True)
states.head()

X = X.drop('State', axis=1)
X.head()

X = pd.concat([X, states], axis=1)
X.head()

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Model training

```python
model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred
```

```python
y_test
```

# Analyze

```python
score = r2_score(y_test, y_pred)
print("Score:", score)
```

```python
sns.distplot(y_pred - y_test)
```
