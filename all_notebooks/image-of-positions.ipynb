```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def nrmlz_x(n, maxn):
    return round((n*1280)/maxn)
def nrmlz_y(n, maxn):
    return round((n*2515)/maxn)
def nrmlz_c(n, maxn):
    return round((n*255)/maxn)

plt.rcParams["figure.figsize"] = (22,31)
plt.rcParams['font.size']=14

x = []
y = []
values = []

newimg = np.ones((2517, 1282), dtype=np.uint8)

f = open("/kaggle/input/autotel-shared-car-locations/2020_02_25.csv", "r")

lines = f.readlines()

for i in range(1, len(lines)):
    line_s = lines[i].split(",")
    x.append(float(line_s[0]))
    y.append(float(line_s[1]))
    values.append(int(line_s[2]))

minx = min(x)
miny = min(y)

for i in range(0, len(x)):
    x[i] -= minx
    y[i] -= miny

maxx = max(x)
maxy = max(y)

for i in range(0, len(y)):
    newimg[nrmlz_y(y[i], maxy)][nrmlz_x(x[i], maxx)] = nrmlz_c(values[i], 24)
    #newimg.putpixel((nrmlz_x(x[i], maxx), nrmlz_y(y[i], maxy)), nrmlz_c(values[i], 24))

plt.imshow(newimg)
plt.show()
```
