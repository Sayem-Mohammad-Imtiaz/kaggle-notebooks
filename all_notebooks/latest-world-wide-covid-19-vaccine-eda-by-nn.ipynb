## In This Exploratory Data Analysis we are exploring and analyzing covid-19 vaccination process progress in countries

## 1- Explore data shape, count, and display a sample of data

```python
df=pd.read_csv('/kaggle/input/latest-worldwide-vaccine-data/Worldwide Vaccine Data.csv')
df.head()
```

## The country with largest number of doses administrered per 100 people

```python
max_doses_index = df['Doses administered per 100 people'].idxmax()
largest= df.iloc[max_doses_index]['Country']
largest
```

## The country with least number of doses administrered per 100 people

```python
min_doses_index = df['Doses administered per 100 people'].idxmin()
least= df.iloc[min_doses_index]['Country']
least
```

## The country with largest % of population fully vaccinated

```python
max_percent_index = df['% of population fully vaccinated'].idxmax()
largest= df.iloc[max_percent_index]['Country']
largest
```

### The country with least % of population fully vaccinated

```python
min_percent_index = df['% of population fully vaccinated'].idxmin()
least= df.iloc[min_percent_index]['Country']
least
```

### Top 10 countries with Total doses administered per 100 people

```python
df_top10=df.nlargest(10, 'Doses administered per 100 people')
df_top10.plot(x = "Country", y = "Doses administered per 100 people", kind = "bar")
plt.show()
```

### Top 10 countries with % of population vaccinated (partially or fully)

```python
df_top10=df.nlargest(10, '% of population vaccinated')
df_top10.plot(x = "Country", y = "% of population vaccinated", kind = "bar")
plt.show()
```
