```python
from fastai.imports import *
from fastai import *
from fastai.vision import *
from torchvision.models import *
```

```python
tfms = get_transforms(do_flip = True, max_lighting = 0.2, max_zoom= 1.1, max_warp = 0.15, max_rotate = 45)
data = ImageDataBunch.from_csv('/kaggle/input/jantahackcomputervision/', folder= 'train_SOaYf6m/images', 
                              valid_pct = 0.0,
                              csv_labels = 'train_SOaYf6m/train.csv',
                              ds_tfms = tfms, 
                              fn_col = 'image_names', 
                              label_col = 'emergency_or_not',
                              bs = 16,
                              size = 300).normalize(imagenet_stats)
```

```python
learn = cnn_learner(data, models.resnet101, metrics=accuracy)
```

```python
learn.fit(epochs = 30, lr = 1.5e-4)
```

```python
id_list=list(test.image_names)
label=[]
for iname in id_list:
    img=open_image("/kaggle/input/jantahackcomputervision/train_SOaYf6m/images/"+iname)
    label.append(learn.predict(img)[0])
    if (len(label)%350 == 0):
        print(f'{len(label)} images done!')

sub = pd.read_csv('/kaggle/input/jantahackcomputervision/ss.csv')
sub['emergency_or_not'] = label
sub['emergency_or_not'] = sub['emergency_or_not'].astype('int')
sub.head()
```
