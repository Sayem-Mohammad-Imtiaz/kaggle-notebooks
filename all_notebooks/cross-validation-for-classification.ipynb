```python
data = pd.read_csv('/kaggle/input/cardio-vascular-disease-detection/cardio_train.csv', sep=';')
data['cardio'].value_counts()
```

```python
data.info()
```

```python
y = data['cardio']
X = data.drop('cardio', axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = .66)
X_train.shape
```

```python
y_train.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier()
model = classifier.fit(X_train, y_train)
classifier.score(X_train, y_train)
```

```python
classifier.score(X_test, y_test)
```

```python
from sklearn.model_selection import KFold, cross_val_score
folds = KFold(n_splits = 10)
scores = cross_val_score(classifier, X, y, cv=folds)
print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
for i in range(1,10):
    print("max_depth=", i, end=": ")
    classifier = DecisionTreeClassifier(max_depth=i)
    scores = cross_val_score(classifier, X, y, cv=10)
    print("%0.2f accuracy with a standard deviation of %0.2f" % (scores.mean(), scores.std()))
```

```python
from sklearn.model_selection import cross_validate

classifier = DecisionTreeClassifier(max_depth=3)
cross_validate(classifier, X, y, cv=10)
```

```python
classifier = DecisionTreeClassifier(max_depth=3)
model = classifier.fit(X_train, y_train)
classifier.score(X_train, y_train)
```

```python
classifier.score(X_test, y_test)
```

```python
import matplotlib.pyplot as plt # data visualization

plt.figure()
plt.title("Feature importances")
plt.barh(X.columns, model.feature_importances_, 1)
```

```python
from sklearn import tree
import matplotlib.pyplot as plt # data visualization

plt.figure(figsize=(20,10))

tree.plot_tree(model, feature_names = X.columns, class_names = ['no', 'yes'], label='root') 
```
