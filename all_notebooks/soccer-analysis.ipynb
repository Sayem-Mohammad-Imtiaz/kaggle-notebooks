```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import missingno as msno
import plotly.express as px

import warnings
warnings.simplefilter(action='ignore', category=Warning)

plt.style.use('dark_background')
```

```python
df = pd.read_csv("../input/english-premier-league202021/EPL_20_21.csv")
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(10,7))
msno.dendrogram(df)
plt.grid()
```

*We have no null values*

***Exploratory Data Analysis***

---



---



***Ages***

```python
plt.figure(figsize=(11, 8))
plt.hist(df['Age'])
plt.xlabel("Ages")
plt.ylabel("Density")
plt.grid()
```

***Players Experiences***

```python
plt.figure(figsize=(11, 8))
plt.scatter(df['Age'],df['Matches'])
plt.xlabel("Age")
plt.ylabel("Matches")
plt.title("Experiences of Players")
plt.grid()
```

***Top Goals***

```python
top_goal = px.scatter(df.nlargest(10, 'Goals')[['Name', 'Goals']],
                 x='Name', y='Goals',template="plotly_dark")
top_goal.update_layout(title_text="Top Goals", )
top_goal.show()
```

***Top Assists***

```python
top_assist = px.scatter(df.nlargest(10, 'Assists')[['Name', 'Assists']],
                 x='Name', y='Assists',template="plotly_dark")
top_assist.update_layout(title_text="Top Assists", )
top_assist.show()
```

***Top Passes Attempted***

```python
top_pass = px.scatter(df.nlargest(10, 'Passes_Attempted')[['Name', 'Passes_Attempted']],
                 x='Passes_Attempted', y='Name',template="plotly_dark")
top_pass.update_layout(title_text="Top Passes Attempted", )
top_pass.show()
```

***Top Defender Score***

```python
defender_score = df[df['Position'].str.contains("DF")]
defender_score = defender_score.drop(defender_score.index[defender_score['Goals'] == 0 ])

defender = px.scatter(defender_score, x='Goals', y='Name', template="plotly_dark")

defender.update_layout(title_text="Top Defender Score", )
defender.show()
```

***Top Yellow Card Owners***

```python
yellow_card = px.violin(df,x='Club',y='Yellow_Cards', template="plotly_dark")
yellow_card.update_layout(title_text="Top Yellow Card Owners")
yellow_card.show()
```

***Top Red Card Owners***

```python
red_card = px.violin(df,x='Club',y='Red_Cards', template="plotly_dark")
red_card.update_layout(title_text="Top Red Card Owners")
red_card.show()
```

***Goals by Teams***

```python
goal_team = df['Goals'].groupby(df['Club']).sum().sort_values(ascending=False).to_frame()
goal_team = px.scatter(goal_team,x='Goals',y=goal_team.index, template="plotly_dark")
goal_team.update_layout(title_text="Goals by Teams")
goal_team.show()
```

***Correlation***

```python
plt.figure(figsize=(18,12))
correlation = sb.heatmap(df.corr(), annot=True, linewidths=1 ,cmap='gist_gray')
correlation.set_title("Correlation of Features", fontdict={'fontsize': 16})
```
