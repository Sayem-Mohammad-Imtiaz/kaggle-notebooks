```python
yalisangxinxi='/kaggle/input/arizona-houses-2021/AZhousingData.csv'
yalisangshuju=pd.read_csv(yalisangxinxi)
print(yalisangshuju.columns)

yalisangshuju=yalisangshuju.dropna(axis=0)

y=yalisangshuju.Price
```

设置路径  文件变量  读取文件 查看columns属性 删除缺失列表行 设置变量y是鸭梨房价


```python
yalitiqu=['beds','baths','sqft']

# ValueError: could not convert string to float: 'Phoenix'
# ValueError: could not convert string to float: 'Apache County'
X=yalisangshuju[yalitiqu]
print(X)
```

设置变量yalitiqu是需要提取的数据 设置X变量引用这个数据

```python
X.describe()
```

通过调用describe方法，我们能得到刚才指定的几个特征的信息

```python
from sklearn.tree import DecisionTreeRegressor
yalimoxing=DecisionTreeRegressor(random_state=1)
yalimoxing.fit(X,y)
```

从sklearn库的tree类中导入 DecisionTreeRegressor(决策树累加器)方法
定义模型。为参数 随机_状态 指定一个数字，以确保每次运行的结果相同
拟合模型

```python
# yalijiage=yalimoxing.
print(yalimoxing.predict(X.head()))#预测出的结果是价格
```

以上是进行简单预测进行出的结果

```python
# 导入平均误差方法
from sklearn.metrics import mean_absolute_error
#创建predicted_home_prices变量，承载了墨尔本数据集中通过X特性得到的房价预测值
predicted_home_prices = yalimoxing.predict(X)
# 调用mean_absolute_error方法，将目标值（原始房价值）y和刚才得到的预测房价值交给该函数进行对比会得到平均误差值
mean_absolute_error(y, predicted_home_prices)
```

```python
from sklearn.model_selection import train_test_split

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 20)
# Define model
yalimoxing = DecisionTreeRegressor()
# Fit model
yalimoxing.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = yalimoxing.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```
