### INFORMATIONS

```python
YEAR_INDEX_LIST = []
```

#### FRP MEAN

```python
FRP_MEAN_LIST = []
```

```python
i_count = 2000
for x_indexing in YEAR_LIST:
    print(str(i_count) + " " + " FRP MEAN")
    YEAR_INDEX_LIST.append(i_count)
    print("---"*10)
    print("\n")
    print(x_indexing["frp"].mean())
    print("---"*20)
    print("\n")
    FRP_MEAN_LIST.append(x_indexing["frp"].mean())
    i_count += 1
```

#### BRIGHTNESS MEAN

```python
BRIGHTNESS_MEAN_LIST = []
```

```python
i_count = 2000
for x_indexing in YEAR_LIST:
    print(str(i_count) + " " + " BRIGHTNESS MEAN")
    print("---"*10)
    print("\n")
    print(x_indexing["brightness"].mean())
    print("---"*20)
    print("\n")
    BRIGHTNESS_MEAN_LIST.append(x_indexing["brightness"].mean())
    i_count += 1
```

#### BRIGHT_T31 MEAN

```python
BRIGHT_T31_MEAN_LIST = []
```

```python
i_count = 2000
for x_indexing in YEAR_LIST:
    print(str(i_count) + " " + " KELVIN MEAN")
    print("---"*10)
    print("\n")
    print(x_indexing["bright_t31"].mean())
    print("---"*20)
    print("\n")
    BRIGHT_T31_MEAN_LIST.append(x_indexing["bright_t31"].mean())
    i_count += 1
```

#### FRP MAX

```python
FRP_MAX_LIST= []
```

#### CONCAT VARIABLES

```python
YEAR_INDEX_LIST = pd.Series(YEAR_INDEX_LIST,name="YEAR")
FRP_MEAN_LIST = pd.Series(FRP_MEAN_LIST,name="FRP_MEAN")
BRIGHTNESS_MEAN_LIST = pd.Series(BRIGHTNESS_MEAN_LIST,name="BRIGHTNESS_MEAN")
BRIGHT_T31_MEAN_LIST = pd.Series(BRIGHT_T31_MEAN_LIST,name="KELVIN_MEAN")


FRP_MAX_LIST = pd.Series(FRP_MAX_LIST,name="FRP_MAX")
BRIGHTNESS_MAX_LIST = pd.Series(BRIGHTNESS_MAX_LIST,name="BRIGHTNESS_MAX")
KELVIN_MAX_LIST = pd.Series(KELVIN_MAX_LIST,name="KELVIN_MAX")
```

```python
MAIN_VARIABLES = pd.concat([YEAR_INDEX_LIST,FRP_MEAN_LIST,BRIGHTNESS_MEAN_LIST,BRIGHT_T31_MEAN_LIST,
                           FRP_MAX_LIST,BRIGHTNESS_MAX_LIST,KELVIN_MAX_LIST],axis=1)
```
