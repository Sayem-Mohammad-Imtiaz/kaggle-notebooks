# DATA ANALYSIS-SUICIDES


*1.SUICIDES IN OVER TIME*


```python
data.plot(kind="scatter",x="year",y="suicides_no",alpha=0.5,color="red",figsize=(10,10),grid=True,label="Suicides No - Year")
plt.xlabel("YEAR")
plt.ylabel("SUICIDES NO")
plt.legend()
plt.title("Suicides in Over Time")
plt.show()
```

*2.SUICIDES IN COUNTRY*

```python
data.plot(kind="line",x="country",y="suicides_no",linestyle=":",color="green",figsize=(10,10),label="SUICIDE NO",alpha=.7,grid=True)
plt.legend()
plt.xlabel("Country")
plt.ylabel("Suicide NO")
plt.title("Suicide No in Country")
plt.show()
```

*3.SUICIDES-AGE CONNECTION*

```python
data.plot(kind="line",x="age",y="suicides_no",label="SUICIDES NO",grid=True,alpha=.7,figsize=(10,10))
plt.legend()
plt.xlabel("AGE")
plt.ylabel("SUICIDES NO")
plt.title("SUICIDES-AGE CONNECTION")
plt.show()
```

*4.SUICIDES-GENERATION CONNECTION*

```python
print(data.generation.unique())
generation_x = data[data.generation == "Generation X"]
silent = data[data.generation == "Silent"]
gi_generation = data[data.generation == "G.I. Generation"]
boomers= data[data.generation == "Boomers"]
millenials= data[data.generation == "Millenials"]
generation_z = data[data.generation == "Generation Z"]
```

```python
f,ax=plt.subplots(figsize=(10,10))
plt.plot(silent.generation,silent.suicides_no,color="red",label="SILENT")
plt.plot(generation_x.generation,generation_x.suicides_no,color="blue",label=("GENERATION X"))
plt.plot(gi_generation.generation,gi_generation.suicides_no,color="green",label=("GI GENERATION"))
plt.plot(boomers.generation,boomers.suicides_no,color="yellow",label=("BOOMERS"))
plt.plot(millenials.generation,millenials.suicides_no,color="orange",label=("MILLENIALS"))
plt.plot(generation_z.generation,generation_z.suicides_no,color="black",label=("GENERATION Z"))
plt.legend()
plt.xlabel("GENERATION")
plt.ylabel("SUICIDES NO")
plt.title("SUICIDES-GENERATION CONNECTION")
plt.show()
```

As you can see,we compered the suicide rate of generations and generation of BOOMERS seems a bit troubled.
