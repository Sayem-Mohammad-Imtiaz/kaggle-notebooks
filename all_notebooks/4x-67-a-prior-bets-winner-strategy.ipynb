## Como funciona?
A estrag√©gia consiste nas seguintes premissas:
- Voc√™ quadriplica sua aposta sempre que perder
- Os eventos tem que ter cota√ß√£o acima de 2.0(para que o retorno seja maior do que 100% do valor apostado).
- N√£o fazer aporte em eventos simut√¢neos
- Suas chances a priori s√£o de 67% de chance de ganhar.


### Exemplifica√ß√£o
considere participar de at√© 3 eventos com uma cota√ß√£o m√°xima de 16 reais, seu aporte no 1¬∞ evento ser√° de 2 reais. 
Nesse cen√°rio, se voc√™ perder 3 eventos seguidos, o seu 3¬∞ evento vai ter um aporte de 16 reais, considere a cota√ß√£o do evento de 2.5, logo 2.5*16=40.0 reais, assim seu retorno nesse evento vencedor √© de 40.0 reais, at√© ent√£o voc√™ perdeu 2+4+16=22 reais,  40-22=18 reais de lucro, voc√™ acaba de ganhar 45% da sua banca no pior cen√°rio!

```python
import numpy as np
total_de_eventos_realizados = []
def simular_banca(valores_aposta, odds_low_value, odds_high_value, print_info=False):
    odds_aposta = np.random.uniform(low=odds_low_value, high=odds_high_value, size=(1,))
    
    valor_investido = np.sum(valores_aposta)
    valor_investido_parcial = []
    evento = ['jogador ganhou', 'perdeu']
    count_ganho=[]
    count_perda=[]
    cum_valor_perdido = []
    for i in range(len(valores_aposta)):
        resultado_evento = np.random.choice(evento, p=[2/3, 1/3])
        if print_info:
            print("Resultado do evento ", resultado_evento)
        total_de_eventos_realizados.append(1)
            
        if resultado_evento=='perdeu':
            count_perda.append(1)
            valor_investido -= valores_aposta[i]
            valor_investido_parcial.append( valores_aposta[i])
            cum_valor_perdido.append(valores_aposta[i])
            if print_info:
                print("VALOR INVESTIDO ", valores_aposta[i])
                print("--------------------------")
            if valores_aposta[i] != 1024:
                odds = valores_aposta[i+1]
            else:
                odds = valores_aposta[1]
        if resultado_evento=='jogador ganhou':
            count_ganho.append(1)  
            valor_investido += (valores_aposta[i]*odds_aposta)
            valor_investido_parcial.append( valores_aposta[i])
            valor_maximo_investido = valores_aposta[i]
            print("Valor m√°ximo investido at√© um evento vencedor ", valor_maximo_investido)
            print("Odds do jogo no evento vencedor ", odds_aposta)
            
            if print_info:
                print("VALOR INVESTIDO ", valores_aposta[i], " ,LUCRO ATUAL A BANCA ",valor_investido[0]- np.sum(valores_aposta))
                print("+++++++++++++++++++++++++++")
            break
    
    valor_investido_total = np.sum(valor_investido_parcial)          
    return valor_investido- np.sum(valores_aposta), valor_investido_total, total_de_eventos_realizados
```
