```python
data["win_by_run"] = np.where(data["result"]=="runs",1,0)
data["win_by_wicket"] = np.where(data["result"]=="wickets",1,0)
```

```python
data["win_by_run"].value_counts()
```

```python
team = data["team1"].unique().tolist()
```

```python
teams1 = []
wins1 = []
for t in team:
    rs_df = data.loc[data['winner'] == t] 
    wins1.append(rs_df["win_by_run"].value_counts())
    teams1.append(t)
```

```python
teams2 = []
wins2 = []
for t in team:
    rs_df = data.loc[data['winner'] == t] 
    wins2.append(rs_df["win_by_wicket"].value_counts())
    teams2.append(t)
```

```python
wins_by_runs = []
for i in range(len(wins1)):
    wins_by_runs.append(wins1[i][1])
```

```python
wins_by_wickets = []
for i in range(len(wins2)):
    wins_by_wickets.append(wins2[i][1])
```

## Wins while Defending.

```python
#Plotting bar chart for number of matches succesfully defended by team
plt.figure(figsize=(10,7))
plt.title("Wins while Defending",fontsize=20)
plt.bar(teams1,wins_by_runs,color=["r","darkred","dodgerblue","b","indigo","orchid","darkorange","gold","salmon","grey","peru","purple"])
plt.xlabel("Teams",fontsize=15)
plt.ylabel("No of wins",fontsize=15)
plt.show()
```

#### MI , CSK and SRH defended their targets which means they have good bowling attack.

## Plotting pie chart for head to head win % by each team.

# CSK

```python
#CSK
z = 1
c = 1
colors = ["r","darkred","dodgerblue","b","indigo","orchid","darkorange","salmon","grey"]
x = 0
plt.figure(figsize=(20,12))
for y in tt:
    if y!="CSK":
        inx = data.loc[((data["team1"] == "CSK") & (data["team2"] == y)) | ((data["team1"] == y) & (data["team2"] == "CSK")),"winner"]
        if len(inx) > 0:
            a = 0
            b = 0
            for i in inx.values:
                if i == "CSK":
                    a = a+1
                else:
                    b = b+1
            if a>=b:
                a = "yellow"
                b = colors[x%10]
                labels = ["CSK",y]
            else:
                a,b = colors[x%10],"yellow"
                labels = [y,"CSK"]
            ax1 = plt.subplot(3,5,c)
            inx.value_counts().plot(kind="pie",autopct='%1.0f%%',colors = [a,b],labels=labels)
            plt.title("CSK VS {}".format(y))
            centre_circle = plt.Circle((0,0),0.70,fc='black')
            fig = plt.gcf()
            fig.gca().add_artist(centre_circle)
            ax1.axis('equal')  
            plt.tight_layout()
                #plt.show()
            z = z+1
            c = c+1
            x = x+1    
```
