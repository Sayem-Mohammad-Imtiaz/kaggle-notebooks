### Dependent and Independent Variables

```python
y = dataset.price.values
X = dataset.drop(['price','model'], axis = 1).values
```

```python
print(X)
print('----->')
print(y)
```

### Encode both transmission and fuelType using OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(),[1,3])], remainder = 'passthrough')
X = np.array(ct.fit_transform(X))
print(X)
```

```python
X[0]
```

### Splitting into train and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

### Train Model + Cross Validation

```python
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(random_state = 42)
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)


from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(reg, X_train, y_train, cv = 10)
print('Accuracy: {:.2f}%'.format(accuracies.mean()*100))
print('Std: {:.2f}%'.format(accuracies.std()*100))

from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor(n_estimators=10, random_state=42)
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)


from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(reg, X_train, y_train, cv = 10)
print('Accuracy: {:.2f}%'.format(accuracies.mean()*100))
print('Std: {:.2f}%'.format(accuracies.std()*100))

from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
from xgboost import XGBRegressor
reg = XGBRegressor()
reg.fit(X_train, y_train)
y_pred = reg.predict(X_test)


from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(reg, X_train, y_train, cv = 10)
print('Accuracy: {:.2f}%'.format(accuracies.mean()*100))
print('Std: {:.2f}%'.format(accuracies.std()*100))

from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
