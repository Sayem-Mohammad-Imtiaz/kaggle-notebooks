```python
import numpy as np
import pandas as pd

ds = pd.read_csv("../input/perry-rhodan-raumschiffe/perry_ds.csv")
```

**prepare data:**

```python
ds=ds.dropna()

qq1=[]
for i in ds.Quellen:
    zw = [int(s) for s in i.split() if s.isdigit()]
    if zw==[]:
        qq1.append([0.])
    else:
        qq1.append(np.mean(zw))
qq2=[]
for i in ds.Quellen_Typ:
    zw = [int(s) for s in i.split() if s.isdigit()]
    if zw==[]:
        qq2.append([0.])
    else:
        qq2.append(np.mean(zw))
    

ds.loc[:,"Quellen"]=np.hstack(qq1).astype(int)
ds.loc[:,"Quellen_Typ"]=np.hstack(qq2).astype(int)
```

**run knn-model**, set 'n_neighbors' to desired no. of neighbors:

```python
from sklearn.neighbors import NearestNeighbors

X = ds[["Größe","Quellen","Quellen_Typ"]].values

nbrs = NearestNeighbors(n_neighbors=10, algorithm='ball_tree').fit(X)
nn_matrix = nbrs.kneighbors_graph(X).toarray()
```
