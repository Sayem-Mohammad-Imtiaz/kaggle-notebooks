```python
# Imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# Read Data
df = pd.read_csv('../input/gdp-annual-growth-for-each-country-1960-2020/GDP_annual_growth.csv')
df.head()
```

**Heatmap for Null values**

```python
sns.heatmap(df.isnull(), cmap = 'viridis')
```

**Filling Null Values with 0**

```python
# df = df.fillna(df.mean(axis=1))
df = df.fillna(0)
df.head()
```

```python
df.nunique()
```

**Cleaning Countries Column**

```python
import pycountry
code=[x.alpha_3 for x in pycountry.countries]
results=df['Country Code'].isin(code)
ls=[]
for i in range(len(results)):
    if(results[i]==False):
        ls.append(i)
len(ls)
```

```python
df=df.drop(ls,axis=0)
df.count()
```

**TOP 10 Couuntries of World wrt GDP in 1999**

```python
plt.figure(figsize=(17, 7))
gdp1999 = df[['Country Name','1999']].sort_values(by='1999', ascending=False)
plt.xticks(rotation=60)
sns.barplot(x='Country Name', y='1999', data=gdp1999.head(10))
```

**TOP 10 Couuntries of World wrt GDP 2010**

```python
plt.figure(figsize=(17, 7))
gdp2010 = df[['Country Name','2010']].sort_values(by='2010', ascending=False)
plt.xticks(rotation=60)
sns.barplot(x='Country Name', y='2010', data=gdp2010.head(10))
```

**TOP 10 Couuntries of World wrt GDP 2019**

```python
plt.figure(figsize=(17, 7))
gdp2019 = df[['Country Name','2019']].sort_values(by='2019', ascending=False)
plt.xticks(rotation=60)
sns.barplot(x='Country Name', y='2019', data=gdp2019.head(10))
```

```python
def plot(country): 
    dfCountry = df.loc[df['Country Name']== country]
    x = np.arange(1960,2020).tolist()
    y = dfCountry.iloc[:,4:64].values.tolist()
    sns.lineplot(x, y[0], linewidth=2)
```

```python
plt.figure(figsize=(17, 7))
plot('United States')
plot('China')
plot('Japan')
plot('Germany')

plt.legend(['United States', 'China', 'Japan', 'Germany'])
```

**Yearby Comperison of Top 4 Countries**

```python
plt.figure(figsize=(17, 7))
plot('Turkey')
plot('Saudi Arabia')
plot('United Arab Emirates')
plot('Pakistan')

plt.legend(['Turkey', 'Saudi Arabia', 'United Arab Emirates', 'Pakistan'])
```

**Comparison of 4 Islamic Countries**

```python
plt.figure(figsize=(17, 7))
plot('Pakistan')
plot('Bangladesh')
plot('Sri Lanka')

plt.legend(['Pakistan', 'Bangladesh', 'Sri Lanka'])
```

**Comparison of 3 Asian Countries**

```python
plt.figure(figsize=(17, 7))
plot('Canada')
plot('Brazil')
plot('India')
plot('United Kingdom')
plt.legend(['Canada', 'Brazil', 'India', 'United Kingdom'])
```
