DATA CLEANING


```python
sns.boxplot(x = 'Pclass', y = 'Age', data = train)
```

```python
#Removing the desprepencies from age column
def age_impute(col):
    Age = col[0]
    Pclass = col[1]

    if pd.isnull(Age):

        if Pclass == 1:
            return 37

        elif Pclass == 2:
            return 29

        elif Pclass == 3:
            return 24
    else:
        return Age
```

```python
train['Age'] = train[['Age', 'Pclass']].apply(age_impute, axis =1)
```

```python
sns.heatmap(train.isnull(),yticklabels=False)
```

```python
#Dropping the column "Cabin"
train_1=train.drop("Cabin",axis=1)
```

```python
sns.heatmap(train_1.isnull(),yticklabels=False)
```

```python
#countplot for visualizing deaths wrt age
sns.countplot(x=train["Survived"],hue=train["Age"]>=35)
```
