# Checking Correlation between features

```python
for i in train.columns:
    for j in train.columns:
        correlation_matrix = train[i].corr(train[j]).round(2)
        if correlation_matrix > 0.7:
            if(i != j):
                print(i,j,correlation_matrix)

        elif correlation_matrix < -0.7:
            if(i !=j):
                print(i,j,correlation_matrix)
```
