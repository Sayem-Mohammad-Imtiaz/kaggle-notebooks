The data provides list of countries with Global Entrepreneurship Index and Women Entrepreneurship Index in 2015. Lets go through the data and see what we can infer from the data.

```python
#loading the dataset.
data=pd.read_csv("/kaggle/input/women-entrepreneurship-and-labor-force/Dataset3.csv",delimiter=";")
data.head(30)
```

```python
#importing the required libraries
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data.shape
```

```python
#Dropping the columns "No" and "Currency" as they add no meaning to the data
data.drop(["No","Currency"],axis=1,inplace=True)
data.shape
```

```python
#Checking the variables datatypes 
data.info()
```

```python
data["Level of development"].value_counts(normalize=True)
```

```python
data["European Union Membership"].value_counts(normalize=True)
```

```python
#Plotting the WEI with Membership and devolopment level nations.
plt.figure(figsize=(12,8))
plt.subplot(2,1,1)
sns.boxplot(x="European Union Membership",y="Women Entrepreneurship Index",data=data)
plt.subplot(2,1,2)
sns.boxplot(x="Level of development",y="Women Entrepreneurship Index",data=data)
plt.show()
```

```python
data1 = data.groupby(['Level of development','European Union Membership'], as_index=False)['Entrepreneurship Index'].mean()
print(data1)
```

```python
sns.violinplot(x ='Level of development', y ='Women Entrepreneurship Index', data = data, hue ='European Union Membership', split = True) 
plt.show()
```

Interestingly, we found that WEI is relatively more in the nations who are not a part of European Union Membership 

Let us see how the inflation rate of a country influences the WEI in the next step.

```python
data['Inflation rate'].value_counts()
```

Creating new features "Inflation","Entrepreneurship" based on binning of existing columns with labels as below.

```python
data['Inflation'] = pd.cut(data['Inflation rate'], [-5,-0.1,0,5,10,30], labels=['negative',"zero", 'low', 'medium',"high"])
data['Entrepreneurship'] = pd.cut(data['Entrepreneurship Index'], [30,45,60,75], labels=['poor', 'good', 'high'])
data.head()
```

```python
#Plotting WEI with Inflation 
plt.figure(figsize=(10,6))
sns.boxplot(x="Inflation",y="Women Entrepreneurship Index",data=data)
plt.show()
```

```python
#Plotting WEI with Entrepreneurship 
plt.figure(figsize=(10,6))
sns.boxplot(x="Entrepreneurship",y="Women Entrepreneurship Index",data=data)
plt.show()
```

```python
#Statistical data of the numerical columns.
data.describe()
```

```python
sns.pairplot(data,vars=["Women Entrepreneurship Index","Entrepreneurship Index","Inflation rate","Female Labor Force Participation Rate"])
plt.show
```

> From the Pairplots, we can say that the countries having high Female Labour Force has the high WEI 

```python
#plotting heatmap to find the correlation among the numeric variables.
plt.figure(figsize = (16, 10))
sns.heatmap(data.corr(), annot = True, cmap="YlGnBu")
plt.show()
```
