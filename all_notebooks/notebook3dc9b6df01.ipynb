```python
df = pd.read_csv("../input/up-crime-data-year-2019/MR data - Compiled Data Set.csv")
df.head()
```

```python

df.rename(columns = {'Murder Happened with reason:\n1:Property/Land Disputes,\n2:Family Dispute\n3:Petty Quarrels,\n4:Money Disputes, \n5:Personal Vendetta, \n6:Love Affairs, \n7:Casteism\n8: Unknown/other':'Murder_reason'}, inplace = True)
df.rename(columns = {'Crime Against Women ( Combined ):\n1. Murder with Rape \n2. Dowry Deaths(Sec. 3048)\n3. Suicide(sec 305/306)\n4. Kidnapping(All)\n5. Acid Attack(Sec. 326A IPC)\n6. Cruelty by Husband/in-laws((Sec.498 A IPC)\n7. Rape only(Sec. 376 or 511 IPC)\n8. Assault on Women with Intent to Outrage her Modesty (Sec. 354 IPC)\n9. Cyber Crimes against Women\n10. Protection of Children from Sexual Offences Act':'Crime_Women'}, inplace = True)

df.tail()
```

```python
df=df.fillna(0)
```

```python
df=df.replace(to_replace ="Property Disputes",value =1)
df=df.replace(to_replace ="Family Dispute",value =2)
df=df.replace(to_replace ="Petty Quarrels",value =3)
df=df.replace(to_replace ="Money Disputes",value =4)
df=df.replace(to_replace ="Personal Vendetta",value =5)
df=df.replace(to_replace ="Love Affairs",value =6)
df=df.replace(to_replace ="Casteism",value =7)
df=df.replace(to_replace ="Unknown reasons",value =8)
df.head()
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
count1=0
count2=0
count3=0
count4=0
for i in range(len(df)) :
    if df.iloc[i, 0]=="Ghaziabad":
        count1=count1+df.iloc[i,4]+df.iloc[i,5]+df.iloc[i,6]
    elif df.iloc[i, 0]=="Kanpur":
        count2=count2+df.iloc[i,4]+df.iloc[i,5]+df.iloc[i,6]
    elif df.iloc[i, 0]=="Lucknow":
        count3=count3+df.iloc[i,4]+df.iloc[i,5]+df.iloc[i,6]
    else:
        count4=count4+df.iloc[i,4]+df.iloc[i,5]+df.iloc[i,6]
city=list(("Ghaziabad","Kanpur","Lucknow","Other"))     
count=list((count1,count2,count3,count4))  
plt.bar(city, count, color ='red',width = 0.4)
 
plt.xlabel("CITY")
plt.ylabel("TOTAL MURDER VICTIM")
plt.title("MURDER")
plt.show()
```

```python
victim=[0,0,0,0,0,0,0,0,0]
#print(type(df.iloc[1,3]))
for i in range(len(df)) :
    if df.iloc[i,3]!=0 :
        victim[int(df.iloc[i, 3])]=victim[int(df.iloc[i, 3])]+(df.iloc[i,4])+df.iloc[i,5]+df.iloc[i,6]
reason=list(("0","1","2","3","4","5","6","7","8"))     
 
plt.bar(reason, victim, color ='red',width = 0.4)
 
plt.xlabel("Reason")
plt.ylabel("TOTAL MURDER VICTIM")
plt.title("MURDER Per REASON")
plt.show()
```
