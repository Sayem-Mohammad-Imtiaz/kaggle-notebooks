The code below convert the timestamp(it must be a int) to normal human readable date i.e, YYYY-MM-DD

```python
import datetime

def timestamp_to_date(timestamp):
    timestamp=int(timestamp) #This step is optional, this is only if the timestamp is given in str format.
    normal_date = datetime.datetime.fromtimestamp(timestamp).strftime('%Y-%m-%d')
    return normal_date
```
