## Prediction

```python
from sklearn.model_selection import train_test_split
x = data.drop('output',axis=1)
y = pd.DataFrame(data,columns=['output'])
x.head()
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

## Logistic Regression as L

```python
from sklearn.linear_model import LogisticRegression
L = LogisticRegression()
```

```python
L.fit(x_train,y_train)
```

```python
r_sq = L.score(x_train,y_train)
print('Coef of determination:',r_sq)
```

```python
pred = L.predict(x_test)
L.score(x_test,y_test)
```

```python
print("Actual output:\n",list(y_test['output']))
print("Predicted output:\n",list(pred))
```

```python
print(*[list(pred)[i] == list(y_test['output'])[i] for i in range(len(list(pred)))], sep=' ')
```

## SGD Classifier as SGD

```python
from sklearn.linear_model import SGDClassifier
SGD = SGDClassifier()
```

```python
SGD.fit(x_train,y_train)
```

```python
r_sq = SGD.score(x_train,y_train)
print("coef of determination:", r_sq)
```

```python
pred = SGD.predict(x_test)
SGD.score(x_test,y_test)
```

```python
print('Actual output:\n',list(y_test['output']))
print('predicted output:\n',list(pred))
```

```python
print(*[list(pred)[i] == list(y_test['output'])[i] for i in range(len(list(pred)))], sep=' ')
```

## K-nearest Neighbours as KNN

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=3)
```

```python
KNN.fit(x_train,y_train)
```

```python
r_sq = KNN.score(x_train,y_train)
print('coef of determination:',r_sq)
```

```python
pred = KNN.predict(x_test)
KNN.score(x_test,y_test)
```

```python
print('Actual output:\n',list(y_test['output']))
print('Predicted output:\n',list(pred))
```

```python
print(*[list(pred)[i] == list(y_test['output'])[i] for i in range(len(list(pred)))], sep=' ')
```
