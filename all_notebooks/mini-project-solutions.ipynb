### Problem 4

```python
col = 'balance'
q1 = bank[col].quantile(0.25)
q2 = bank[col].quantile(0.5)
q3 = bank[col].quantile(0.75)
iqr = q3 - q1
lw = q1 - 1.5 * iqr
uw = q3 + 1.5 * iqr
outliers = bank[(bank[col]<lw) | (bank[col] > uw)][col]
perc_outliers = outliers.shape[0] / bank.shape[0] * 100
print('Percentage of outliers in %s column is: %.2f %%' % (col, perc_outliers))
bank[col].plot.box();
```
