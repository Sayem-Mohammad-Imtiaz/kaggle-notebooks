# Finding similar tweets with KNN

```python
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm

ds = pd.read_csv("../input/emotion-detection-from-text/tweet_emotions.csv")
```

```python
#data prep

a_words = pd.Series(np.hstack(ds.content.str.split(" ").values))
wl = a_words.value_counts()[(a_words.value_counts()>100).values].index
zz = pd.Series(wl)

val = []
for i in tqdm(ds.content.str.split(" ")):
    val.append(zz.isin(i).values)
    
X = np.squeeze(val).astype(int)

y = ds.sentiment.values
```

```python
#learn neighbor matrix

from sklearn.neighbors import KNeighborsClassifier

nbrs = KNeighborsClassifier(n_neighbors=10, algorithm='ball_tree').fit(X,y)
nn_matrix = nbrs.kneighbors_graph(X).toarray()
```

```python
#display neighbors of a tweet

numb = 1000
print("Neighbors of "+str(ds.iloc[numb].tweet_id))
ds[nn_matrix[numb]==1.]
```
