# Pre-processing

As we know the data is categorical we have to initiate with some tasks here -
            1. Checking leakages
            2. Filling the leakages  (if present)
            3. If the data is producing any combination or sequential formation then reshuffling the data by rows
            4. Numerating the categorical data

* Leakages

```python
for j in range(23):
    cnt=0
    for i in range(26):
        cnt+=len(df[df[colm[j]]==chr(i+97)])
    print('no. of leakages in',colm[j],'is :',len(df)-cnt)
```

data is leakage free. 

So we have enough data to analyze the model and predict.


* Data reshuffling

It is best to shuffle data always. Cause chances of getting a combination after this process are close to zero.

```python
df
```

```python
df=df.sample(frac=1)
df
```

After shuffling we have to make the data set numeric to initiate further process.

* Numeration of categorical data

```python
def non_num_data(df):             # function to turn all categorical data into numbers
    columns=df.columns.values
    
    for column in columns:
        text_digit_vals={}
        def con_to_int(val):
            return text_digit_vals[val]
        
        if df[column].dtype!=np.int64 and df[column].dtype!=np.float64:
            column_contents=df[column].values.tolist()
            unique_elements=set(column_contents)
            x=0
            for unique in unique_elements:
                if unique not in text_digit_vals:
                    text_digit_vals[unique]=x
                    x+=1
            df[column]=list(map(con_to_int,df[column]))
    return df
non_num_data(df)
```

# Prediction with RandomForestClassifier

As the data has been turned numric we can start prediction

Creating X and Y

```python
X_df=df.drop('class',1)
y_df=df['class']
```

* *Train data* and *Test data* preparation

```python
X_train,X_test,y_train,y_test=train_test_split(X_df,y_df,test_size=0.2,random_state=0)
print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

Creating model

```python
model=RandomForestClassifier()
model
```

Fitting model

```python
model.fit(X_train,y_train)
```

Creating *prediction* data

```python
y_pred=model.predict(X_test)
y_pred
```

Checking *Accuracy*

```python

accuracy_score(y_test,y_pred)
```

WOW ! That RandomForestClassifier makes an perfect accuracy.
