```python
# This is to impport the csv file,assign texts to columns and start indexing from 1 
bankmarketing=pd.read_csv('../input/bank-marketing-campaign/bank-additional/bank-additional-full.csv',header=0,sep=';')
bankmarketing.index = bankmarketing.index + 1
bankmarketing.head(3)
```

```python
# This is to extract only numeric columns in preparation for correlation analysis 
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']

numbankmarketing = bankmarketing.select_dtypes(include=numerics)
print(numbankmarketing )
```

```python
# Analysing  the variables two at a time to determine correlation 

Age_duration = bankmarketing[['age', 'duration']]
cAge_duration = Age_duration.corr()

Age_campaign= bankmarketing[['age', 'campaign']]
cAge_campaign = Age_campaign.corr()

Age_pdays= bankmarketing[['age', 'pdays']]
cAge_pdays = Age_pdays.corr()

Age_previous= bankmarketing[['age', 'previous']]
cAge_previous = Age_previous.corr()
                             
Age_empvarrate= bankmarketing[['age', 'emp.var.rate']]
cAge_empvarrate = Age_empvarrate.corr()

Age_conspriceidx= bankmarketing[['age','cons.price.idx']]
cAge_conspriceidx = Age_conspriceidx.corr()

Age_euribor3m= bankmarketing[['age','euribor3m']]
cAge_euribor3m = Age_euribor3m.corr()

Age_nremployed = bankmarketing[['age','nr.employed']]
cAge_nremployed  = Age_nremployed.corr()


print(cAge_duration,"\n","\n",cAge_campaign,"\n","\n",cAge_pdays,"\n","\n",cAge_previous,"\n","\n",cAge_empvarrate,"\n","\n",cAge_conspriceidx,"\n","\n",cAge_euribor3m,"\n","\n",cAge_nremployed)
```

```python
# Tabulated summary of the summary of findings in a N X N matrix
correlation_summary = numbankmarketing.corr()
limit_float = round(correlation_summary, 2)
print(limit_float)
# to display as a table
fig, ax = plt.subplots()
fig.patch.set_visible(True)
ax.axis('off')
ax.axis('tight')
ax.table(cellText=limit_float.values, colLabels=limit_float.columns, loc='center')
fig.tight_layout()
plt.show()
```
