```python
from sklearn.feature_extraction.text import TfidfVectorizer


tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')
```

```python
matrix = tfv.fit_transform(movies['Overview'])
```

```python
matrix
```

```python
matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel
sig = sigmoid_kernel(matrix, matrix)
```

```python
sig[0]
```

```python
indices = pd.Series(movies.index, index=movies['Series_Title']).drop_duplicates()
```

```python
indices
```

```python
indices["Giant"]
```

```python
sig[996]
```

```python
list(enumerate(sig[indices['Giant']]))
```

```python
def give_rec(title, sig=sig):
    id = indices[title]
    scores = list(enumerate(sig[id]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    scores = scores[1:11]
    series_indices = [i[0] for i in scores]
    return pd.DataFrame({"title":movies['Series_Title'].iloc[series_indices],"scores":scores})
```

```python
give_rec('The Dark Knight')
```
