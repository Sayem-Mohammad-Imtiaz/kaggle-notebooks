```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor as rgs
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

Input 2 DATA FILES USD TO PKR 2

```python
df = pd.read_csv('../input/usd-pk5-20192021/usd_pkr.csv')
```

```python
df1 = pd.read_csv('../input/usd-pk5-20192021/usd_pkr_2.csv')
```

# concat 1999-2019 and 2019-2021

# Cleaning data

```python
dfnew = pd.concat([df1 ,df])
```

```python
dfnew.isnull().sum()
```

```python
dfnew.dropna(how ='any')
```

```python


dfnew['Date'] = pd.date_range(start='15-May-2021', end='07-Jun-1999'  , periods =  dfnew.shape[0])

```

```python
dfnew['Date'] = pd.to_datetime(df.Date)
```

```python
dfnew.dtypes
```

```python
x_ir = dfnew['Change %'].astype(str)
x_ir = x_ir.replace({'%':''}, regex = True)
dfnew['Change %'] = x_ir
```

```python
y = dfnew.Price
```

```python
dfnew.dtypes
```

# Feature Engineering--> 3 Methods

# HeatMapping


```python
features = [ 'Open' , 'High' , 'Low' ,'Change %'  ]
```

```python
dfnew_X = dfnew[features] 
```

```python
y = dfnew.Price
```

```python
y
```

```python
corrmat = dfnew.corr()

top_corr_features = corrmat.index
plt.figure(figsize=(10,10))
g=sns.heatmap(dfnew[top_corr_features].corr(),annot=True,cmap="RdYlGn")

```

# Random Forest ML Algorithm


```python
X_train, X_test, Y_train, Y_test = train_test_split(dfnew_X,y)
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
forest_model = RandomForestRegressor(random_state=1)
```

```python
forest_model.fit(X_train, Y_train)
```

```python
predicts = forest_model.predict(X_test)
```

```python
predicts
```

```python
print(mean_absolute_error(Y_test, predicts))
```

```python
mean_squared_error(Y_test, predicts)
```

```python
fig, ax = plt.subplots()
ax.scatter(Y_test, predicts)
ax.plot([Y_test.min(), Y_test.max()], [Y_test.min(), Y_test.max()], 'k--', lw=2 )
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')

plt.show()
```
