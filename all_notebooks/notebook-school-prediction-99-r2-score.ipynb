```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
scaler = MinMaxScaler()
df_cleaned[["n_student","pretest", "school_type_id","teaching_method_id","gender_id","lunch_id","school_setting_id"]] = scaler.fit_transform(df_cleaned[["n_student","pretest", "school_type_id","teaching_method_id","gender_id","lunch_id","school_setting_id"]])
```

```python
df_cleaned.describe()
```

```python
X=df_cleaned
X.drop(columns=['posttest'])
y=df_cleaned["posttest"]
```

```python
X.describe()
y.describe()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gbrt = GradientBoostingRegressor()
gbrt.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
y_pred = gbrt.predict(X_test)
errors = mean_squared_error(y_pred,y_test)
print(errors)
tree_rmse = np.sqrt(errors)
print(tree_rmse)
```
