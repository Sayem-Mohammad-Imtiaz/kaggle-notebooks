```python
nlp = spacy.load('en')
```

```python
def normalize(msg):
    
    doc = nlp(msg)
    res=[]
    
    for token in doc:
        if(token.is_stop or token.is_digit or token.is_punct or not(token.is_oov)):
            pass
        else:
            res.append(token.lemma_.lower())
    
    return " ".join(res)
```

```python
df['Clean_plot'] = df['Plot'].apply(lambda x: normalize(x))
df.sample(5)
```
