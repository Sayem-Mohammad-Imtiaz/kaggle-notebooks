```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../input/ipl-complete-dataset-20082020/IPL Matches 2008-2020.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(14,8))
sns.set_style('whitegrid')
sns.countplot(x=df.winner)
plt.xticks(rotation=90)
```

We can clearly see that Mumbai Indians have won a large number of matches.
Teams like Deccan Chargers, Pune Warriors, Kochi Tuskers Kerala, Rising Pune Supergaints and Delhi Capitals have relatively less wins because they are new teams or older teams which have played very less seasons of IPL

```python
plt.figure(figsize=(14,8))
sns.countplot(x=df.city)
plt.xticks(rotation=90)
```

Most of the matches are played in India, but we can see Dubai and Cape Town because due to Corona, IPL was shifted to Dubai and due to elections, IPL was shifted to South Africa.

```python
plt.figure(figsize=(14,8))
sns.countplot(x=df.umpire1)
plt.xticks(rotation=90)
```

HDPK Dharmsena is usually prefered for umpiring as he has umpired most of the matches compared to others.

```python
plt.figure(figsize=(14,8))
sns.countplot(df.umpire2)
plt.xticks(rotation=90)
```

S Ravi has been Second Umpire in most of the IPL matches.

```python
plt.figure(figsize=(14,8))
sns.countplot(df.toss_decision)
```

Most of the times the team won toss, has decided to field.

```python
plt.figure(figsize=(14,8))
sns.countplot(df.toss_winner)
plt.xticks(rotation=90)
```

Mumbai Indians has won the toss most of the times, followed by Chennai Super Kings and Kolkata Knight Riders.

```python
plt.figure(figsize=(14,8))
sns.countplot(df.eliminator)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(14,8))
sns.countplot(df.result)
```

We can say from the above graph that the team which batted in second innigs or the team which chased had more probability to win.

```python
plt.figure(figsize=(14,10))
sns.scatterplot(x=df.team1,y=df.team2,hue=df.toss_winner)
plt.xticks(rotation=90)
```

Here we can compare that when team1 and team 1 are face to face, then who is the toss winner more frequently.

```python
plt.figure(figsize=(14,10))
sns.scatterplot(x=df.venue,y=df.winner,hue=df.toss_winner,alpha=0.5)
plt.xticks(rotation=90)
```

Here we can see that there is a high probability for a team to win a toss on its home ground and also win the game.
