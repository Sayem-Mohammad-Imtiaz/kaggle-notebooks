```python
BASE_PATH = '/kaggle/input/game-data-of-warau-arsnotoria/'
sigil_table_df = pd.read_csv(BASE_PATH + 'sigil_table.csv', index_col=0)
char_table_df = pd.read_csv(BASE_PATH + 'char_table.csv', index_col=0)
```

```python
# データ更新の確認
display(sigil_table_df[sigil_table_df['No'] == sigil_table_df['No'].max()])
display(char_table_df[char_table_df['No'] == char_table_df['No'].max()])
```

# シジルの統計

```python
sigil_table_df['タイプ']\
.value_counts()\
.sort_index()\
.plot\
.pie(colors=["g", "y", "r", "b", "m"], title="シジルのタイプ割合", autopct="%1.1f%%", ylabel="")
```

```python
data = sigil_table_df['レア度'].value_counts().sort_index()
data = pd.DataFrame(data)
plt.title("シジルのレア度別の個数", fontsize=16)
plt.bar(data.index, data['レア度'], color="b", width=0.7, align="center")
for x, y in zip(data.index, data['レア度']):
    plt.text(x, y, y, ha='center', va='bottom') # 棒グラフ内に数値を書く
plt.show()
```

# キャラの統計

```python
char_table_df['ギフト']\
.value_counts()\
.sort_index()\
.plot\
.pie(colors=["y", "gray", "b", "m", "r", "g", "orange"], title="キャラのギフト割合" ,autopct="%1.1f%%", ylabel="")
```

```python
char_table_df['属性']\
.value_counts()\
.sort_index()\
.plot\
.pie(colors=["y", "b", "m", "r", "g"], title="キャラの属性割合" ,autopct="%1.1f%%", ylabel="")
```

```python
sns.pairplot(data=char_table_df, hue='ギフト', palette=["r", "orange", "m", "y", "b", "g", "gray"])
plt.show()
```
