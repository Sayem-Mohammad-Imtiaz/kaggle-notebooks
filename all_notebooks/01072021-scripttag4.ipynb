# Recap

```python
# Lesen der Daten
data = pd.read_csv("/kaggle/input/cusersmarildownloadsearningcsv/earning.csv", delimiter = ";")

data.info()
```

```python
data.head()
```

```python

#DATA CLEANING
# Erstelle eine Liste mit allen relevanten Spalten
berufe = []
for group in data.columns:
    # Entfernt Spalten die keine wichtigen Informationen für uns haben
    if group == "males":
        pass
    elif group == "persons":
        pass
    elif list(data.columns).index(group) > 1:
        berufe.append(group)

#berufe

# Umbenennung der Spalten
label_berufe = ['managers',
                'professionals',
                'technicians',
                'serviceworkers',
                'administration',
                'sales',
                'operators',
                'labourers']
label_gender = ['_f', '_m', '_p']

# Erstellen eines Aufgeräumten DataFrames
new_columns = []
for gender in label_gender:
    for beruf in label_berufe:
        new_columns.append(beruf + gender)
        
new_data = data[berufe]
#new_data.info()

new_data.columns = new_columns
#new_data

# Ergänzt das Jahr als Index
new_data.index = data["year"]
new_data.head()
```

```python
# Mean und Std sind nicht sonderlich nützlich für uns,
# wenn wir uns für eine zeitliche Entwicklung interessieren.

new_data["managers_f"].describe()
```

```python
# Wir haben uns die Lohnentwicklung einer bestimmten Berufsgruppe angesehen
plt.plot(new_data["managers_f"], color="red", label="managers_f")
plt.plot(new_data["managers_m"], color="blue", label="managers_m")

plt.legend()
plt.title("Lohnentwicklung zwischen Männer und Frauen")
plt.ylabel("Mean Stundenlohn in US$")
plt.xlabel("Jahr")

plt.show()
```

```python
# Zudem haben wir die Lohndifferenz beider Geschlechter geplottet

plt.plot(new_data["managers_m"] - new_data["managers_f"], label="managers")
plt.plot(new_data["operators_m"] - new_data["operators_f"], label="operators")

# Nulllinie
plt.plot(new_data.index, np.zeros(len(new_data.index)), color="red", linestyle="--")

plt.legend()
plt.xlim(2004, 2017)
plt.style.use("seaborn")
plt.title("Lohndifferenz zw Mann und Frau")
plt.ylabel("Lohndifferenz in US$")
plt.xlabel("Jahre")
plt.show()
```

$$a_{pers} = p \cdot a_{male} + (1 - p) \cdot a_{female}$$

Wir suchen *p*! 

$$p = \frac{a_{pers} - a_{female}}{a_{male} - a_{female}}$$

```python
new_data.info()
```

```python
try:
    a_pers = np.array(new_data.iloc[year, 16:])
    a_male = np.array(new_data.iloc[year, 8:16])
    a_female = np.array(new_data.iloc[year, 0:8])
    anteil =  (a_pers - a_female) / (a_male - a_female)
    
except:
    print("ZERO DEVISION!")
    
anteil
```

```python
for i in range(len(label_berufe)):
    plt.bar(label_berufe[i], anteil[i])


plt.plot(range(-1, len(label_berufe)+1), [0.5]*(len(label_berufe)+2), color="red", linestyle="--")

plt.title(f"Prozentualer Anteil von Frauen {new_data.index[year]}")
plt.ylabel("Anteil an Frauen")
plt.xlim(0 - 0.5, len(label_berufe)-0.5)

plt.xticks(rotation = 90)
plt.show()
```

```python
x = np.random.randn(20, 1) + 10
y = x*2 + np.random.randn(20, 1)

data = np.array([x, y])
data
```

```python
data.shape
```

```python
data = data.reshape(2, 20)
data
```

```python
df = pd.DataFrame(data.T, columns=["x", "t"])
df.head()
```

```python
plt.scatter(df["x"], df["t"])
plt.show()
```

```python
x0 = np.arange(7, 13)
y0 = x0*2

plt.scatter(df["x"], df["t"])
plt.plot(x0, y0, "k--")
plt.legend(["$y =2 \cdot x$", "points"])
plt.show()
```

```python
df["r"] = np.abs(np.random.randn(20)) * 20
df.head()
```

```python
plt.scatter(df["x"], df["t"], color="r")
plt.scatter(df["r"], df["t"], color="g")
plt.plot(x0, y0, "k--")
plt.legend(["$y =2 \cdot x$", "points", "random"], loc="lower right")
plt.show()
```

```python
df.corr()
```

```python
import seaborn as sns
plt.figure()
sns.heatmap(df.corr(), annot=True, cmap='jet')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(np.array(df["x"]), np.array(df["t"]), test_size = 0.2)
```

```python
try:
    model.fit(X_train, y_train)
except Exception as error:
    print(error)
```

```python
# Mehrere Features

x1 = np.random.randn(20, 1) + 10
x2 = np.random.randn(20, 1) + 10
y = x1*2 - x2*3 + np.random.randn(20, 1)
rand = np.abs(np.random.randn(20)) * 20
rand = rand.reshape(20, 1)
data = np.array([x1, x2, rand,  y]).T
data = data.reshape(20, 4)
```

```python
df = pd.DataFrame(data.reshape(20, 4), columns=["x1", "x2", "random", "target"])
df
```

```python
plt.scatter(df["x2"], df["target"])
plt.show()
```

```python
import seaborn as sns
plt.figure()
sns.heatmap(df.corr(), annot=True, cmap='jet')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(np.array(df[["x1", "x2"]]), np.array(df["target"]), test_size = 0.2)
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
from sklearn import linear_model

model = linear_model.LinearRegression()

model.fit(X_train, y_train)
model.score(X_test, y_test)
```
