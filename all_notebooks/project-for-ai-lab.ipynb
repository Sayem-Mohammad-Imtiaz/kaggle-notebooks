```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
y_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
classifier = Sequential()

classifier.add(Dense(output_dim = 40, init = 'uniform', activation = 'relu', input_dim = len(X[0])))
classifier.add(Dense(output_dim = 25, init = 'uniform', activation = 'relu'))
classifier.add(Dense(output_dim = 32, init = 'uniform', activation = 'sigmoid'))
classifier.add(Dense(output_dim = 25, init = 'uniform', activation = 'relu'))
classifier.add(Dense(output_dim = 1, init = 'uniform', activation = 'sigmoid'))

classifier.compile(optimizer = 'sgd', loss = 'mean_squared_error', metrics = ['accuracy'])
classifier.fit(X_train, y_train, batch_size = 10, nb_epoch = 10)
```

```python
y_pred = classifier.predict(X_test)

y_pred = (y_pred > 0.5)
y_test = (y_test > 0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
print(cm)
labels = ['Negative' , 'Positive']
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(cm)
plt.title('Confusion matrix of the classifier')
fig.colorbar(cax)
ax.set_xticklabels([''] + labels)
ax.set_yticklabels([''] + labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
