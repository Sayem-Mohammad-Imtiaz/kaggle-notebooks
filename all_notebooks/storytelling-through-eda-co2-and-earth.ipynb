```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 
df=pd.read_csv('../input/global-co2-emissions/CO2Emission_LifeExp.csv')
```

# Global Carbon Dioxide Emissions EDA And Visualizations

```python
df.drop('Code',axis=1,inplace=True)
```

```python
df 
```

# Exploratory Data Analysis

```python
df.describe()
```

```python
df.info()
```

```python
dfsort1=df.sort_values(by='CO2Emissions',ascending=False)
```

## Top 10 Countries With Highest Carbon Dioxide Emissions(Absolute)

```python
dfsort1.head(10)
```

```python
sns.catplot(x="Country", y="CO2Emissions", kind="bar", data=dfsort1.head(10),height=5, aspect=27/9,)
```

**Inferences:**
* The Geographical Distribution Of This list Is Quite Diverse, But It's Clear That Asia Is A Big Producer Of CO2
* Most Countries In The List Are Large Economies
* Most Of These Countries Also Have The Largest Populations In The World
* Most Of These Countries Have Population Life Expectancy On Par With The Global Average.

```python
dfsort2=df.sort_values(by='Percapita',ascending=False)
```

## Top 10 Countries With Highest Carbon Dioxide Emissions(Per Capita)

```python
dfsort2.head(10)
```

```python
sns.catplot(x="Country", y="Percapita", kind="bar", data=dfsort2.head(10),height=5, aspect=28/9,)
```

**Inferences:**
* The Geographical Distribution Of This list Is Somewhat Diverse, But It's Clear That MiddleEast Is A Big Producer Of CO2 Per Capita.
* Most Of These Countries Are Rich And Low Populated.
* All These Countries Have Population Life Expectancy Higher Than Global Average.

## Top 10 Countries With Lowest Carbon Dioxide Emissions(Absolute)

```python
dfsort3=df.sort_values(by='CO2Emissions')
dfsort3.head(10)
```

```python
sns.catplot(x="Country", y="CO2Emissions", kind="bar", data=dfsort3.head(10),height=5, aspect=27/9,)
```

**Inferences:**
* All Of These Countries Have Extremely Low Population
* Most Of These Are Small Islands Or Coastal Nations That Do Not Comprise Of Any Large Cities.
* Some Of The Emissions Are Less Than That Of Cities Like New York, Mumbai And London.

## Top 10 Countries With Lowest Carbon Dioxide Emissions(Per Capita)

```python
dfsort4=df.sort_values(by='Percapita')
dfsort4.head(10)
```

```python
sns.catplot(x="Country", y="Percapita", kind="bar", data=dfsort4.head(10),height=5, aspect=20/9,)
```

**Inferences:**
* All Of These Countries Have Very Low Population
* Many Of These Countries Are Located In Africa And Nearby Regions
* The Life Expectancy Of Most Of These Countries Are Much Below World Average.
* Most Of These Countries Have Bad Economic Indicators.

## Top 10 Countries With Highest Reduction In Carbon Dioxide Emissions As Compared To Previous Year(Good Performers)

```python
d1=df.sort_values(by="YearlyChange",ascending=True)
d2=df.sort_values(by="YearlyChange",ascending=False)
```

```python
d1.head(10)
```

```python
sns.catplot(x="Country", y="YearlyChange", kind="bar", data=d1.head(10),height=5, aspect=27/9,)
```

**Inferences:**
* Half Of The Countries Are European.
* Brazil Is Known For The Famous Amazon Forests And Such Statistics Are Not Surprising. Some Even Call It The Lungs Of The Earth.
* Africa And South America Have A Share Of Two Countries Each.
* No Country From Asia.
* Most Countries Except Brazil And United Kingdom Are Mediocre In Economic Prowess.
* Even Though Countries Like UK, Greece, Portugal and Trinidad & Tobago Have Performed Well Over The Last Year, They Still Have Above Average Per Capita CO2 Emission. 

## Top 10 Countries With Highest Increase In Carbon Dioxide Emissions As Compared To Previous Year(Bad Performers)

```python
d2.head(10)
```

```python
sns.catplot(x="Country", y="YearlyChange", kind="bar", data=d2.head(10),height=5, aspect=27/9,)
```

**Inferences:**
* Most Of These Countries Are Asian.
* Almost All Countries Have Bad Socio-Economic Indicators especially GDP and Life Expectancy.
* Almost All Countries Have Very Good Per Capita Emission Statistics.
* It Is Surprising To See That Countries With Bad Per Capita Performed Far Better Than The Ones With Low Per Capita Emissions.
* Most Of These Countries Have Significant Populations.
