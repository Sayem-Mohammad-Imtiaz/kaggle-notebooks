```python
from sklearn.preprocessing import LabelEncoder
import numpy as np
dfcc=np.zeros(dfc.shape).astype(int)
le=[]
for i in range(dfc.shape[1]):
    if not isinstance(dfc.iloc[:,i],str):
        le.append(LabelEncoder())
        dfcc[:,i]=le[-1].fit_transform(dfc.iloc[:,i])
        
        
```

```python
dfcc=pd.DataFrame(dfcc,columns=dfc.columns)
cormat=dfcc.corr()
cormat=np.triu(cormat,k=1).reshape(-1,12)
new=pd.DataFrame(cormat,columns=dfc.columns)
sns.heatmap(new,annot=True,fmt='.2f',yticklabels=new.columns[::-1])
```
