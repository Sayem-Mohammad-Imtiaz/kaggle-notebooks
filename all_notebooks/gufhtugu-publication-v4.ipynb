# 1. Top Ten Books of guftughu-publication?

We have a name of books but the issuse is that the multiple book names in a order are join together with the backslash. We refine the book name column and split the book name in order to find the total unique books value.

```python
#splitting the book name by '/'
book_Name = gufh_data['Book Name'].str.split('/')

#creating a new columns and add to 3 index
gufh_data.insert(3,'Books_Name',book_Name,True)
gufh_data.head()
```

```python
#Thanks to @Usman Ayaz
#books name with the total selling count
books = gufh_data.Books_Name.explode().value_counts()
book_topTen = dict(books[:10])

k = list(book_topTen.keys())

#printing top ten books name
print('Top Ten Best Selling Books of Ghuftughu Publication\n')

for i in range(len(book_topTen)):
    print(i+1," ", k[i])
```

```python
print('The Top selling book of Ghuftughu-Publication is: ', k[0])
```

```python
v = list(book_topTen.values())

#visualization of top ten selling book
plt.figure()
sns.barplot(x = k , y = v)

plt.title("Top Ten Selling Books of Ghuftughu-Publication ")
plt.xlabel('Books Name')
#thanks to Usman Ayaz
plt.xticks(rotation=95, size=10)
plt.yticks(size=10)
plt.ylabel('Selling Count')

plt.show()
```

# 2. Co-relation wtih Order Status 

```python
#dropping nan value from dataset
gufh_data = gufh_data.dropna(axis=0, inplace=False)
```

```python
gufh_data.head()
```

```python
#calculate the correlation of different variable with Order Status
correlation = gufh_data.groupby(['Order Status']).corr()
sns.heatmap(correlation)

plt.title("Correlation of Order Status with other variables")
plt.xlabel('Other Columns')
plt.xticks(size=10)
plt.yticks(size=10)
plt.ylabel('Order Status')
```

Note: see the above right side template for better understanding
* The darkest the negative correlation
* The lightest the positive correlation


```python
print(correlation)
```

```python
#Total number of Orders countdifferent categories
ord_st = gufh_data['Order Status'].value_counts()
print('Order Status Statistics\n', ord_st)
```

```python
plt.figure(figsize=(10, 5))
ord_st.plot(kind='bar',color=['Red','green','blue'])
plt.title("Order Status Category division", size=20)
plt.xlabel('Order Status categories', size=15)
plt.ylabel('Order Value Count', size=15)
plt.show()
```
