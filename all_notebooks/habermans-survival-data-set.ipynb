```python
# Find survival of patients who had undergone surgery for breast cancer
survival.plot(kind = 'kde', x = 'Survival_status', y = 'Age', title = 'kde Plot')
plt.show()
# kind='kde' Kernel Density Estimation.
```

```python
survival.plot(kind = 'scatter', x = 'Survival_status', y = "Age", title = 'Scatter Plot')
plt.show()
```

Age does't seems a major factor regarding survive patient more than 5 years after operation

However 80+ aged patient surely not survived more than 5 years.


```python
sns.boxplot(x = 'Survival_status', y = 'Age', data = survival)
plt.show()
```



*   Min age for the patients survived more than 5 years seems 30 and max age seems 76, and majority of patients age is between 42-60.
*   Min age for the patients survived less than 5 years seems 35 and max age seems 85, and majority of patients age is between 45-61.
*   Numbers are not accurate, just an estimate.








```python
#boxplot for all columns
survival.iloc[:,:].boxplot(figsize=[10,15])
plt.subplots_adjust(bottom=0.50)
plt.show()
```



*   Majority of patients age is between 42-61.

```python
survival.plot(kind = 'kde', x = 'Survival_status', title = 'kde plot', figsize = [10,7])
plt.show()
```

```python
# Violin Plot
sns.violinplot(x = 'Survival_status', y = 'Age', data = survival)
plt.show()
```

```python
# Checking correlation: plotting heatmap
survival['Age'].hist(bins = 20, grid = True)
plt.show()
```

```python
corr_mat = survival.corr()
plt.figure(figsize = [20,12])
sns.heatmap(corr_mat, annot = True, cmap = 'RdYlBu_r')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Checking the data distribution among the columns. 
survival.plot(kind='density', subplots=True, layout=(1,4), sharex=False, legend=True, fontsize=2, figsize=(15,3))
plt.show()
```

By observing Survival_status, we can conclude than more number of the patients survived more than 5 years, less died within 5 years after the operation.
