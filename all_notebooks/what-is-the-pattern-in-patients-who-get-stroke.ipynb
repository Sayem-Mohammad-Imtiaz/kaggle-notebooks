# Parameters analysis of patients who get stroke

### This is simple analysis patients who get stroke base on the parameters given
### Please upvote if you like this notebook

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data=pd.read_csv('/kaggle/input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')
data.head()
```

```python
data.shape
```

```python
data['bmi']=data['bmi'].fillna(data['bmi'].mean())
```

```python
segment = []

for i in data['age'].values:
    if (i < 20):
        segment.append('less than 20')
    if (i >= 20) & (i<30):
        segment.append('between 20 & 30')
    if (i >= 30) & (i<40):
        segment.append('between 30 & 40')
    if (i >= 40) & (i<50):
        segment.append('between 40 & 50')
    if (i >= 50) & (i<60):
        segment.append('between 50 & 60')
    if (i >= 60) & (i<70):    
        segment.append('between 60 & 70')
    if (i >= 70):
        segment.append('More than 70')
```

```python
data['age_segment']=segment
```

```python
segment2 = []

for i in data['avg_glucose_level'].values:
    if (i < 80):
        segment2.append('less than 80')
    if (i >= 80) & (i<130):
        segment2.append('between 80 & 130')
    if (i >= 130) & (i<180):
        segment2.append('between 130 & 180')
    if (i >= 180):
        segment2.append('more than 180')

data['glucose_seg']= segment2
```

```python
segment3 = []

for i in data['hypertension'].values:
    if i == 1:
        segment3.append('hypertension')
    if i ==  0:
        segment3.append('no hypertension')
```

```python
data['hypertension']=segment3
```

```python
segment4 = []

for i in data['heart_disease'].values:
    if (i == 1):
        segment4.append('heart_disease')
    if (i ==  0):
        segment4.append('no heart_disease')
```

```python
data['heart_disease']=segment4
```

```python
segment5 = []

for i in data['stroke'].values:
    if (i == 1):
        segment5.append('stroke')
    if (i ==  0):
        segment5.append('no stroke')
```

```python
data['stroke']=segment5
```

## We have taken all patients who got stroke in one dataset (dfst)

```python
dfst=data[data['stroke']=='stroke']
```

```python
agegrp=data.groupby('stroke')['age'].mean().reset_index()
fig = px.bar(agegrp, y="age", x="stroke",color='stroke', log_y=False)
fig.update_layout(title='Average age of patients got stroke vs not got stroke')
fig.show()
```
