```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
data = None
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        data = pd.read_csv(os.path.join(dirname, filename))
data.drop(columns=['Unnamed: 0','Unnamed: 0.1','registradoEm'],inplace=True)
data['publicadoEm'] =  pd.to_datetime(data['publicadoEm'], infer_datetime_format=True)
data.info()
```

```python
data.pivot_table(index=["categoria","situacao"],values=["ticker"], aggfunc='count')
```

```python
data.pivot_table(index=["categoria"],values=["ticker"], aggfunc='count')
```

```python
data.pivot_table(index=["tipo"],values=["ticker"], aggfunc='count')
```

```python
data.pivot_table(index=["situacao"],values=["ticker"], aggfunc='count')
```

```python
data.pivot_table(index=["especie"],columns=["situacao"],values=["status"], aggfunc=['count'],fill_value=0,margins=True)
```

```python
pd.pivot_table(data,index=["modalidade","analisado","status","situacao"],values=["ticker"],aggfunc=["count"])
```

```python
data.sort_values(by=['publicadoEm'], inplace=True, ascending=True)
data.groupby(pd.Grouper(key='publicadoEm',freq='Y'))["ticker"].count().plot.bar()
```
