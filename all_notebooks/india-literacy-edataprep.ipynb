```python
from dataprep.eda import plot, plot_correlation, create_report, plot_missing
```

```python
df = pd.read_csv('/kaggle/input/india-literacy-data-district-wise/Literacy Data 2011.csv')
df.head()
```

```python
plot(df)
```

#Numerical column, it computes column statistics, and generates a histogram, kde plot, box plot and qq-normal plot:

```python
plot(df, "Literacy")
```

#Categorical column, it computes column statistics, and plots a bar chart and pie chart.

#It has WordCloud, WordFrequency and Word Lenght. Isn't that amazing!

```python
plot(df, "State")
```

#The relationship between two columns with plot(df, x, y)

When x and y are both numerical columns, it generates a scatter plot, hexbin plot and box plot.

```python
plot(df, "Unnamed: 0","Literacy")
```

#When x and y are both categorical columns, it plots a nested bar chart, stacked bar chart and heat map.

```python
plot(df, "District","State")
```

#With one or two specific columns, it provides a more detailed plot for the specific columns by passing column names as the parameter.

```python
plot(df, "Literacy", bins=20)
```

#Correlation matrix between columns.

```python
plot_correlation(df)
```

#Correlation of specified element with all other attributes.

```python
plot_correlation(df, "Literacy")
```

#Analyze missing values

```python
plot_missing(df)
```
