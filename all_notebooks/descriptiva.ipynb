```python
correlacionM = datos.loc[:,['room_type', 'price',
       'minimum_nights', 'number_of_reviews',
       'reviews_per_month', 'calculated_host_listings_count',
       'availability_365']].corr(method='pearson')
#importo libreria grafica
import matplotlib.pyplot as plt
import seaborn as sb
#realizar grafico
sb.heatmap(correlacionM, 
            xticklabels=correlacionM.columns,
            yticklabels=correlacionM.columns,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5)
```

```python
tablaCruzada = pd.crosstab(datos.neighbourhood_group, datos.room_type) 
#realizar grafico
sb.heatmap(tablaCruzada, 
            xticklabels= tablaCruzada.columns,
            yticklabels=tablaCruzada.index,
            cmap='RdBu_r',
            annot=True,
            linewidth=0.5)
```

```python
#inicialmente se importa la liberia de plotly
import plotly.graph_objects as px
```

```python
fig = px.Figure(data=[px.Histogram(x=datos.price,nbinsx=150)])
fig.show()
```

```python
fig = px.Figure(px.Histogram2d(
        x=datos.availability_365,
        y=datos.room_type
    ))
fig.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure(go.Scattermapbox(
    mode='markers',
    lon = datos.longitude, lat = datos.latitude,
    marker = dict(
        size=4,
        color= datos.price, 
        colorscale='Viridis', 
        showscale=True
    )))

fig.update_layout(
    mapbox = {
        'style': "stamen-terrain",
        'center': {'lon': -73.935242, 'lat': 40.730610 },
        'zoom': 10},
    showlegend = False)

fig.show()
```
