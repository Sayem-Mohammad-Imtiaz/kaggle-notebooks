```python
import pandas as pd
import numpy as np
```

Import all modules that are needed, then load the dataset.

```python
df = pd.read_csv('../input/billboard-hot-100-weekly-charts-2020/billboard_top_100.csv')
df.sample()
```

```python
df.info()
```

I decide to delete some columns that unuseful and change the 'date' datatype from object to DateTime.

```python
df.drop(["Unnamed: 0","id"],axis=1, inplace=True)
```

```python
df["date"] = pd.to_datetime(df["date"], format = '%Y-%m-%d')
```

```python
df.agg(['max','min'])['date']
```

I see the newest and the oldest data from this set. As seen, the newest song is released on 2020-12-26 and the oldest song is released on 2020-01-04.

# The Newest Released Song

```python
df[['artist','song','date']].sort_values(by='date').head(1)
```

# The Oldest Released Song

```python
df[['artist','song','date']].sort_values(by='date',ascending=False).head(1)
```

```python
dfnew = df.groupby(['artist', 'song']).mean().round(2).reset_index()
```

```python
dfnew
```

This table shows a list of 5 artists and songs who stays on the chart and how many weeks.

```python
dfnew[['artist','song','weeks_on_chart']].sort_values(by='weeks_on_chart',ascending=False).head()
```

This table shows a list of 5 artists and songs by sorting the rank.

```python
dfnew[['artist','song','rank']].sort_values(by='rank',ascending=False).head()
```

This table shows a list of 5 artists and songs by sorting the rank last week.

```python
dfnew[['artist','song','rank_last_week']].sort_values(by='rank_last_week',ascending=False).head()
```

This table shows a list of 5 artists and songs by sorting the rank this week.

```python
dfnew[['artist','song','peak_rank']].sort_values(by='peak_rank', ascending=False).head()
```
