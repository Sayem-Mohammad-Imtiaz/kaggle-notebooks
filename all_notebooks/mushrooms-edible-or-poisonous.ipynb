## Data Preparation:

#### Cap Shape:

We see, in the visualization above, that Cap Shape has 6 levels. The 'conical' level has only 4 observations, all of which are of the 'poisonous' mushroom class. The 'sunken' level has 23 observations, all of which are of the 'edible' mushroom class. Further examination shows that the 'bell' class is dominated by 'edible' mushrooms and the 'knobbed' class is dominated by 'poisonous' mushrooms. Both of these levels are smaller when compared to the 'convex' and 'flat' levels, which have a more even distribution of poisonous and edible mushrooms. With this in mind, the 'sunken' level will be combined with the 'bell' level and 'conical' level will be combined with the 'knobbed' level in the next cell.

```python
#Creates dictionary to help group cap shapes as described above.
cap_shape_dict = {'x': 'convex', 'f': 'flat', 'k': 'knobbed_conical', 'b': 'bell_sunken',
                  's': 'bell_sunken', 'c': 'knobbed_conical'}

#Creates a series of dummy variables.
cap_shape_dummies = pd.get_dummies(dat['cap-shape'].replace(cap_shape_dict.keys(), 
                                                            cap_shape_dict.values()),
                                                            prefix = 'cap_shape')

#Adds dummy variables to dataset.
dat = pd.concat([dat, cap_shape_dummies], axis=1)

#Drop original cap shape variable.
del dat['cap-shape']

#A similar pattern will be followed in future cells to create and add dummy variables so
#no comments will be made to avoid repetitive statements.
```

#### Cap Surface:

Cap Surface has 4 levels. One of the levels, 'grooves', has only 4 observations and they are all for 'poisonous' mushrooms. This level will be combined with the 'scaly' level, which has the highest amount of poisonous mushrooms by cap surface and it dominated by poisonous mushrooms. 

```python
cap_surface_dict = {'y': 'scaly_grooves', 'g': 'scaly_grooves', 's': 'smooth', 'f': 'fibrous'}

cap_surface_dummies = pd.get_dummies(dat['cap-surface'].replace(cap_surface_dict.keys(),
                                                                cap_surface_dict.values()),
                                                                prefix = 'cap_surface')

dat = pd.concat([dat, cap_surface_dummies], axis=1)

del dat['cap-surface']
```

#### Cap Color:

Cap Color has 10 levels. Half of these levels have 168 or fewer observations. We would like to bin these smaller levels so as to avoid variables with near zero variance. To do this, we will bin smaller variables dominated by poisonous mushrooms and do the same with edible mushrooms cap color levels. 'Pink' and 'Buff' will become one level. 'Purple', 'Cinnamon' and 'Green' will become another. This will leave us with 7 dummy variables rather than 10.

```python
cap_color_dict = {'n': 'brown', 'g': 'gray', 'e': 'red', 'y': 'yellow', 'w': 'white',
                  'b': 'buff_pink', 'p': 'buff_pink', 'c': 'cinn_purp_green', 
                  'r': 'cinn_purp_green', 'u': 'cinn_purp_green'}

cap_color_dummies = pd.get_dummies(dat['cap-color'].replace(cap_color_dict.keys(),
                                                            cap_color_dict.values()),
                                                            prefix = 'cap_color')

dat = pd.concat([dat, cap_color_dummies], axis=1)

del dat['cap-color']
```

#### Bruises:

This variable only tells us whether a mushroom has bruising or not. We will create a dummy variable to shift this information into numeric form in the next cell.

```python
bruises_dict = {'t': 'True', 'f': 'False'}

bruises_dummies = pd.get_dummies(dat['bruises'].replace(bruises_dict.keys(), 
                                                        bruises_dict.values()),
                                                        prefix='bruises')
                                 
dat = pd.concat([dat, bruises_dummies], axis=1)

del dat['bruises']
```

#### Odor:

There are 9 levels for mushroom odor. Most mushrooms in the sample have either a 'foul' or 'none' odor. The rest are mostly dominated by poisonous or edible mushrooms and have much fewer observations. So the levels will be combined as follows: pungent, foul, creosote, fishy, spicy, and musty will be one level, almond and anise will be another and none will remain a category of its own. This will gives us 3 variables rather than 9.

```python
odor_dict = {'n': 'none', 'f': 'fo_pu_cr_fi_sp_mu', 's': 'fo_pu_cr_fi_sp_mu', 
             'y': 'fo_pu_cr_fi_sp_mu', 'a': 'almond_anise', 'l': 'almond_anise',
             'p': 'fo_pu_cr_fi_sp_mu', 'c': 'fo_pu_cr_fi_sp_mu', 
             'm': 'fo_pu_cr_fi_sp_mu'}

odor_dummies = pd.get_dummies(dat['odor'].replace(odor_dict.keys(), odor_dict.values()),
                              prefix='odor')

dat = pd.concat([dat, odor_dummies], axis=1)

del dat['odor']
```

#### Gill Attachment:

The vast majority of mushrooms have an 'free' gill attachment. But we want to keep this variable at least until we see how it performs in a baseline model. On its own, it may not add much information to the model but in combination with other variables, it may show some value. 

```python
gill_attachment_dict = {'f': 'free', 'a': 'attached'}

gill_attachment_dummies = pd.get_dummies(dat['gill-attachment'].replace(gill_attachment_dict.keys(),
                                                                    gill_attachment_dict.values()),
                                                                    prefix="gill_attachment")

dat = pd.concat([dat, gill_attachment_dummies], axis=1)

del dat['gill-attachment']
```

#### Gill Spacing:

Gill spacing shows two levels. Each dominated by one type of mushroom, which makes it look like an attractive variable for helping us determine whether a mushroom is poisonous or not.

```python
gill_spacing_dict = {'c': 'close', 'w': 'crowded'}

gill_spacing_dummies = pd.get_dummies(dat['gill-spacing'].replace(gill_spacing_dict.keys(),
                                                                  gill_spacing_dict.values()),
                                                                  prefix="gill_spacing")

dat = pd.concat([dat, gill_spacing_dummies], axis=1)

del dat['gill-spacing']
```

#### Gill Size:

Much like gill spacing, we see two levels dominated by one or the other class of mushroom. 

```python
gill_size_dict = {'b': 'broad', 'n': 'narrow'}

gill_size_dummies = pd.get_dummies(dat['gill-size'].replace(gill_size_dict.keys(),
                                                            gill_size_dict.values()),
                                                            prefix='gill_size')

dat = pd.concat([dat, gill_size_dummies], axis=1)

del dat['gill-size']
```

#### Gill Color: 

We see a variable with 12 levels and a few of them can be binned into a single level. The bins will be as follows: buff and green, black, brown, gray, pink, white, chocolate, purple, red and orange and finally, yellow.

```python
gill_color_dict = {'b': 'buff_green', 'p': 'pink', 'w': 'white', 'n': 'brown', 
                   'g': 'gray', 'h': 'chocolate', 'u': 'purple', 'k': 'black',
                   'e': 'red_orange', 'y': 'yellow', 'o': 'red_orange', 
                   'r': 'buff_green'}

gill_color_dummies = pd.get_dummies(dat['gill-color'].replace(gill_color_dict.keys(),
                                                              gill_color_dict.values()),
                                                              prefix= 'gill_color')

dat = pd.concat([dat, gill_color_dummies], axis=1)

del dat['gill-color']
```

#### Stalk Shape:

Here we have two levels with each leaning towards toward one of the two classes of mushrooms.

```python
stalk_shape_dict = {'t': 'tapering', 'e': 'enlarging'}

stalk_shape_dummies = pd.get_dummies(dat['stalk-shape'].replace(stalk_shape_dict.keys(),
                                                                stalk_shape_dict.values()),
                                                                prefix= 'stalk_shape')

dat = pd.concat([dat, stalk_shape_dummies], axis=1)

del dat['stalk-shape']
```

#### Stalk Root:

There are 5 levels here. The 'rooted' level is made up entirely of edible mushrooms albeit only a small amount of observations when compared to the other levels. We will combine the 'equal' and 'club' levels as they both show observations dominated by edible mushrooms but do contain some observations where the mushroom is not edible. This will leave us with 4 variables.

#### Stalk Surface Above Ring:

We have 4 levels here but will combine 'smooth', 'fibrous' and 'scaly' which all lean toward 'edible' and leave silky as a stand alone level.

```python
above_ring_dict = {'s': 'sm_fi_sc', 'k': 'sm_fi_sc', 'f': 'sm_fi_sc', 'y': 'silky'}

above_ring_dummies = pd.get_dummies(dat['stalk-surface-above-ring'].replace(above_ring_dict.keys(),
                                                                            above_ring_dict.values()),
                                                                            prefix='above_ring_surface')

dat = pd.concat([dat, above_ring_dummies], axis=1)

del dat['stalk-surface-above-ring']
```

#### Stalk Surface Below Ring:

We will create 2 variables following the same method used in the stalk surface above ring variable.

#### Stalk Color above ring:

#### Stalk Color below ring:

```python
stalk_color_below_dict = {'w': 'white', 'g': 'gray_red_or', 'p': 'pink', 'n': 'brown',
                          'b': 'buff_cinn_yel', 'e': 'gray_red_or', 'o': 'gray_red_or',
                          'c': 'buff_cinn_yel', 'y': 'buff_cinn_yel'}

stalk_color_below_dummies = pd.get_dummies(dat['stalk-color-below-ring'].replace(
                                                            stalk_color_below_dict.keys(),
                                                            stalk_color_below_dict.values()),
                                                            prefix= 'stalk_color_below_ring')

dat = pd.concat([dat, stalk_color_below_dummies], axis=1)

del dat['stalk-color-below-ring']
```

#### Ring Type:

```python
class_dict = {'p': 'poisonous', 'e': 'edible'}

class_dummies = pd.get_dummies(dat['class'].replace(class_dict.keys(),
                                                    class_dict.values()),
                                                    drop_first=True, 
                                                    prefix="class")

dat = pd.concat([dat, class_dummies], axis=1)

del dat['class']
```

## Modeling: 

```python
#First we split the data into training and test sets:
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(dat.iloc[:, 0:79], dat.iloc[:, -1])
```

```python
#Now we will tune the training model parameters for SVC using cross validation:
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.svm import SVC
```

```python
Cs = [0.001, 0.01, 0.1, 1, 10]
gammas = [0.001, 0.01, 0.1, 1]

param_grid = {'C': Cs, 'gamma': gammas}

svc = SVC()
RSCV = RandomizedSearchCV(svc, param_grid, cv=10, n_jobs=-1) 
```

```python
RSCV.fit(X_train, y_train)
```

```python
print(RSCV.best_params_)
print(RSCV.best_score_)
```

```python
mod = SVC(gamma=0.01)
mod.fit(X_train, y_train)
```

```python
preds = mod.predict(X_test)
```

```python
accuracy_score(y_test, preds)
```

Using support vector classification with a gamma of 0.01, a model was fitted using the training set. Once the model was fit, we used the predict method to find the predicted values of y when fitting the trained model on the test predictors. This returned to us an accuracy of 99.85% when compared to the actual values of y in the withheld test data.

```python
confusion_matrix(y_test, preds)
```
