** How many people have the word Chief in their job title? (This is pretty tricky) **

```python
def chief_string(Title):
    if 'chief' in Title.lower():
        return True
    else:
        return False # not necessary to define the function, can input the condition under lambda
```

```python
sum(sal['JobTitle'].apply(lambda Title: chief_string(Title)))
#or
#sal[sal['JobTitle'].str.contains('chief', case=False)].count()['EmployeeName'] # using the contains built-in function
```
