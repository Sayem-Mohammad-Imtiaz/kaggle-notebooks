#### 2- Iport dataset

```python
# # import dataset
# ----------------

dataset = pd.read_csv('/kaggle/input/years-of-experience-and-salary-dataset/Salary_Data.csv')
```

```python
print("NumPy: ", np.__version__)
print("Pandas: ", pd.__version__)
print("Matplotlib: ", pl.__version__)
print("Seaborn: ", sns.__version__)
print("SKLearn: ", sklearn.__version__)
```

```python
X= dataset.iloc[ : , : -1].values
y= dataset.iloc[ : , -1].values
```

```python
X.shape
```

## *Sklearn*

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn .metrics import r2_score
```

#### 3- Spliting The dataset int Training Set and Test Set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,  random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

#### 4 - Training the Simple Linear Regression model on the Training set

```python
model =LinearRegression()
model.fit(X_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

#### 5 - Predicting the Test set results

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

#### 6 - Visualising the Training set results

```python
plt.scatter(X_train, y_train, color ='yellow')
plt.plot(X_train, model.predict(X_train), color = 'purple')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

#### 7 - Visualising the Test set results

```python
plt.scatter(X_test, y_test, color = 'green')
plt.plot(X_train, model.predict(X_train), color = 'blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
model.score(X_test, y_test)
```

```python
r2_score(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
print(r2_score(y_test, y_pred), round(2)* 10, "%")
```
