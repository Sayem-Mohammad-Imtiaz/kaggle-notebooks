Después de hacer un análisis rápido de los gastos de acuerdo a la ocupación de forma más profunda, se encontró que en todas las ocupaciones hay individuos en prácticamente todos los rangos de compra. Lo que varía son las cantidades en algunos rangos, el volumen y su distribución.

Los highlights importantes fueron:
- Si se quiere hacer una estrategia de Marketing, no hay que desechar el alcanzar a la mayoría de los rangos de edad (solo poniendo a un lado los de 0-17 y +55) ya que hay gente con importante poder adquisitivo distribuidos entre ellos.
- La city category va totalmente correlacionado con el nivel de gastos de los usuarios siendo "C" el más alto y "A" el más bajo.
- La mayoría de usuarios de la base pertenecen a la categoría "B".
- Si se quiere hacer una estrategia de Marketing, no hay que desechar el alcanzar a todas las ocupaciones ya que en todos hay gente con importante poder adquisitivo. En especial habría que resaltar ciertas ocupaciones donde se concentra mayor nivel de compras: el 17, 12 y 15.
- El grupo con mayor nivel de gasto son los hombres en el rango de 51-55 años.
- A partir del rango de edad de 55 años baja bastante las compras por lo que es importante dar seguimiento del porque ese grupo cambia tan repentinamente sus hábitos o si es algo generacional.
- Los usuarios no casados del rango de edad +55 son los que cuentan con mayor número de compras mayores superando hasta otros rangos, pero cuando estan casados ese número baja drásticamente. En comparación con el rango 51-55 se ve que tanto casados como no casados tienen un gran número de este tipo de compras.
- La categoría "C" de ciudad es un indicador robusto e independiente en prácticamente todas las comparaciones con las distintas categorias, indicando una fuerta cantidad de compras mayores.
