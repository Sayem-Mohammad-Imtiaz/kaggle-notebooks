```python
X=df[["Height"]].values
y=df[["Weight"]].values
```

```python
X
```

### Spliting up of Dataset into Training and test Dataset.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =train_test_split(X,y,random_state=42)
```

```python
model=LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

# Visualizing TRAIN MODEL 

```python
plt.style.use("seaborn")
plt.scatter(X_train,y_train,c="blue")
plt.plot(X_train,model.predict(X_train), color='red')
plt.title("TRAIN Model")
plt.xlabel('HEIGHT')
plt.ylabel('WEIGHT')
plt.show()
```

# Visualizing TEST MODEL 

```python
plt.scatter(X_test,y_test,c="blue")
plt.plot(X_train,model.predict(X_train), color='red')
plt.title("TEST Model")
plt.xlabel('HEIGHT')
plt.ylabel('WEIGHT')
plt.show()
```

```python
plt.scatter(X_test,y_test,color='blue')
plt.plot(X_test,y_pred,color='red')
plt.title('Test Set Result')
plt.xlabel('Height')

plt.ylabel('Weight')
plt.show()
```

#### Comparing with TEST RESULTS

```python
plt.scatter(X_test,y_test,c="blue")
plt.plot(X_test,y_pred,color='red')
plt.plot(X_train,model.predict(X_train), color='red')
plt.show()
```

### R2 Score 

```python
from sklearn.metrics import r2_score,mean_squared_error
coefficient_of_dermination = r2_score(y_pred, y_test)
print('R2 Score = ',coefficient_of_dermination)
error=mean_squared_error(y_pred, y_test)
print('Error = ',error)
```

## How to calculate BMI(Body Mass Index) of a person !!

```python
df['BMI']=df['Weight']/(df['Height']*df['Height'])
df
```

```python
sns.distplot(df["BMI"],bins=40)
```

BMI American women of age 30–39 COMES UNDER Normal weight (18.5–24.9) [SEE HERE ](https://www.nhlbi.nih.gov/health/educational/lose_wt/BMI/bmicalc.htm)
