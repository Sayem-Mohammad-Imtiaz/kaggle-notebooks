```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px 
import warnings 

warnings.simplefilter(action='ignore')
```

```python
df = pd.read_csv(r'../input/hepatitis-c-dataset/HepatitisCdata.csv')
df.shape
```

```python
data = df.copy()
```

```python
data.info()
```

```python
data.isna().sum()
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
data.head()
```

# EDA

```python
labels = data['Sex'].value_counts().index
values = data['Sex'].value_counts()

fig = px.pie(data, names=labels, values=values, hole=0)
fig.update_layout(title='Gender Distribution', template='plotly_dark', hoverlabel=dict(
    font_size=20
))
fig.update_traces(hovertemplate='%{value}', pull=0.01)
fig.show()
```

```python
plt.figure(figsize=(10, 7))
sns.distplot(data['Age'])
plt.show()
```

```python
data.head()
```

```python
data['Category'].value_counts()
```

```python
labels = data['Category'].value_counts().index
values = data['Category'].value_counts()

fig = px.pie(data, names=labels, values=values, hole=0)
fig.update_layout(title='Category Distribution', template='plotly_dark', hoverlabel=dict(
    font_size=20
))
fig.update_traces(hovertemplate='%{label}: %{value}', pull=0.1)
fig.show()
```

```python
data.head()
```

```python
a = 2  # number of rows
b = 5  # number of columns
c = 1 # initialize plot counter

fig = plt.figure(figsize=(30,10))

for i in data.columns[3:]:
    plt.subplot(a, b, c)
    plt.xlabel(i)
    sns.boxplot(y=data[i])
    c = c + 1

plt.show()
```
