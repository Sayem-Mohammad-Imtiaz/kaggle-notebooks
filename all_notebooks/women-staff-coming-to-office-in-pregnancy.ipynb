```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
print('Libraries imported')
```

```python
#importing data
url='/kaggle/input/fertility-rate-vs-participation-in-workforce/SLR2.csv'
data=pd.read_csv(url)
data.head()
```

```python
#Basic analysis over data
print('Dimensions of dataset:{}'.format(data.shape),'\n','-'*75)
print('Number of duplicated values:{}'.format(data.duplicated().sum()),'\n','-'*75)
print('Rows containing null values:\n{}'.format(data.isnull().sum()),'\n','-'*75)
print('Schema of dataset:\n',data.info(),'\n','-'*75)
print('Descriptive Analysis:\n{}'.format(data.describe()),'\n','-'*75)
print('Describing Correlation of the data:\n{}'.format(data.corr()),'\n','-'*75)
```

**Hence, our data has no errors and is ready to analyse.**

```python
x=data['fertility rate'].to_numpy()
y=data['worker percent'].to_numpy()
x=x.reshape(-1,1)
y=y.reshape(-1,1)
```

**Now let's see if there is any relationship between the fertility rate and worker percent in our data.**

```python
plt.scatter(x,y)
plt.xlabel('Fertility Rate')
plt.ylabel('Worker Percent')
plt.title('Fertility Rate vs Worker Percent')
plt.show()
```

**From the above graph we can clearly see that there is an inverse relationship between the fertility rate and worker percent.**

**Now lets use this data and make a machine learning model for further predictions.**

```python
#Creating train, test splits
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#creating model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
print('Model Trained')
```

**Now lets see if our model represents data correctly or not.**

```python
line=x*model.coef_+model.intercept_
plt.plot(x,line)
plt.scatter(x,y)
plt.xlabel('Fertility Rate')
plt.ylabel('Worker Percent')
plt.title('Fertility Rate vs Worker Percent')
plt.show()
```

**From here we can see that our model represents the data correctly and could be used for further predictions.**

**Now lets check that how much do the predicted values differ from actual values.**

```python
y_pred=model.predict(x_test)
y_pred
```

```python
df=pd.DataFrame(data={'Actual value':y_test.flatten(),'Predicted value':y_pred.flatten()})
df
```

**From here we can observe that the predicted value is a bit lesser than the actual value. 
So, to correct that, we will now calculate Mean Absolute Error(MAE)**

```python
from sklearn.metrics import mean_absolute_error
mae=mean_absolute_error(y_pred,y_test)
print('The mean absolute error is ',mae)
```

**Since the predicted value is lesser than the actual value so, to get the right prediction, we will have to add MAE to the predicted values, in future predictions.**

**Now lets calculate the worker percent for women having fertility rate of 2.711304**

```python
target=[[3.71]]
prediction=model.predict(target)
print("The predicted worker percentage of women having 2.711304 fertility rate is",prediction[0,0])
```

**Now we need to add the MAE to the prediction so as to get the correct value.**

```python
#Adding MAE to the prediction
prediction[0,0]=prediction[0,0]+mae
print('Hence the corrected prediction is',prediction[0,0])
```

**To get the correct prediction, we need to add MAE everytime we make a prediction.**
