```python
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
df = pd.read_csv('/kaggle/input/cargurus-1/cars.csv')
```

```python
df.head(1)
```

```python
#Example of how to grab the raw price for a certain make
df[df['Title'].str.lower().str.contains('ford')]['Raw Price'].mean()
```

```python
#Start a make list which we'll populate later
make_list = []
```

```python
#Makes directly from the csv
title_list = df['Title']
```

```python
#Clean up the car titles to only include the make (ex: Ford Fusion -> Ford)
for title in title_list:
    make_list.append(title.split(' ')[1].split(' ')[0].lower())
```

```python
#Function that returns the mean(avg) of a given make.
def get_manf_mean(make : str):
    try:
        return(df[df['Title'].str.lower().str.contains(f'{make}')]['Raw Price'].mean())
    except:
        print('nope')
        return None
```

```python
#First make in make_list
make_list[0]
```

```python
#This list will be populated with dictionaries containing makes and their average prices
make_avg_df_list = []
```

```python
#Average price in the cars csv.
get_manf_mean('')
```

```python
#Populate the dictionary list
for make in tqdm(make_list):
    make_avg_df_list.append({
        "Make" : make,
        "Avg $" : get_manf_mean(make)
    })
```

```python
#Convert list of dictionaries to a Pandas DataFrame
df1 = pd.DataFrame(make_avg_df_list)
```

```python
vals = df1['Make'].value_counts().to_frame()
```
