```python
nRowsRead = 1000 # specify 'None' if want to read whole file
df = pd.read_csv('../input/cusersmarildownloadsbquarantinecsv/bquarantine.csv', delimiter=';', encoding = "ISO-8859-1", nrows = nRowsRead)
df.dataframeName = 'bquarantine.csv'
nRow, nCol = df.shape
print(f'There are {nRow} rows and {nCol} columns')
df.head()
```

```python
fig,ax=plt.subplots(figsize=(18,6))
sns.set_style('dark')
plt.subplot(1, 2, 1)
g1=sns.countplot(df['Division'],palette='rocket', hue = df['Geometry'],**{'hatch':'/','linewidth':3})
g1.set_title("Division")

plt.subplot(1, 2, 2)
g2=sns.countplot(df['Geometry'],palette='rocket',hue= df['Geometry'],**{'hatch':'/','linewidth':3})
g2.set_title("Geometry")
```

```python
plt.figure(figsize=(10,5))
sns.countplot(y="Division",data=df,
             facecolor=(0,0,0,0),linewidth=5,
             edgecolor=sns.color_palette("RdYlGn"))
plt.title('Division Quarantine for Covid-19',
         fontsize=15);
```

```python
import plotly.express as px
fig = px.bar(df, x= "Division", y= "total_quarantine", color_discrete_sequence=['crimson'], title="Bangladesh Quarantine for Covid-19")
fig.show()
```

```python
fig = px.bar(df, x= "Distname", y= "Shape_Area", color_discrete_sequence=['cyan'], title="Bangladesh Quarantine for Covid-19")
fig.show()
```

```python
fig = px.treemap(df, path=['Division'], values='total_quarantine',
                  color='total_quarantine', hover_data=['Division'],title="Bangladesh Divisions Quarantine for Covid-19",
                  color_continuous_scale='RdYlGn')
fig.show()
```
