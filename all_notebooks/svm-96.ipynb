```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn import preprocessing
```

```python
#data = pd.read_csv('train.csv')
data = pd.read_csv('../input/mobile-price-classification/train.csv')
print(data.head().T)
```

```python
print(data.describe().T)
```

```python
# Check missing data:
print(data.isna().sum())
```

```python
data.hist(bins=30, figsize=(15, 15))
```

```python
x = data.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
data_scaled = pd.DataFrame(x_scaled)

data_scaled.boxplot(figsize=(15, 15))
```

```python
# Find important features with correlation:
print(data.corr()['price_range'].sort_values())
```

```python
# Drop dependent variable and low correlations:
lowCorrs = list(data.corr()['price_range'].sort_values().keys()[:16]); print("Drop:", lowCorrs)

X = data.drop(['price_range', *lowCorrs], axis=1);print(X.shape)
y = data['price_range'];print(y.shape)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=22, shuffle=True)
```

```python
# Fit model via support vector machine:
model_svm = SVC()
model_svm.fit(X_train, y_train)
model_svm.score(X_test, y_test)
```

```python
# Confusion matrix:
y_pred = model_svm.predict(X_test)
CM = confusion_matrix(y_test, y_pred)
print(CM)
```
