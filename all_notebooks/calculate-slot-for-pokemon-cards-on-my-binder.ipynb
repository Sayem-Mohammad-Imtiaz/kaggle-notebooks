```python
pkdf = pd.read_csv("../input/pokemon.csv")
```

```python
pkdf.columns
```

```python

name_number = pkdf[['name','pokedex_number','generation']]


# A pasta tem N paginas de 18 slots cada.
# queremos calcular para cada pokemon em ordem da pokedex seu slot na pasta, com o par (página,slot_na_pagina)
# na virada entre gerações pular uma pagina, todas as gerações começam numa pagina nova

slots_per_page = 18

name_number['page_number'] = np.around(np.ceil(name_number['pokedex_number'].values / slots_per_page),decimals = 0).astype(int)
name_number['slot_number'] = ((name_number['pokedex_number'].values - 1) % slots_per_page ) + 1
```

```python
# pad dataframe so we have placeholder pokemons to fill the gap between generations

gens = list(set(name_number.generation.values))

for gen in gens:
    # get last pokemon from generation slot (we want to push the next generation to a new page)
    slot = name_number[name_number.generation == gen].tail(1).slot_number
    placeholder_num = 18 - slot
    # add to the pokedex number of subsequent pokemon so the numbers work out
    name_number.loc[name_number.generation > gen,'pokedex_number'] = name_number.loc[name_number.generation > gen,'pokedex_number'].apply(lambda x : x + placeholder_num).values
    # new slot values for this offset
    name_number['slot_number'] = ((name_number['pokedex_number'].values - 1) % slots_per_page ) + 1


    
# recalculate with correct offsets
name_number['page_number'] = np.around(np.ceil(name_number['pokedex_number'].values / slots_per_page),decimals = 0).astype(int)

```
