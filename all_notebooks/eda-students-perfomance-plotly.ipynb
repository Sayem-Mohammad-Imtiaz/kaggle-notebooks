```python
from plotly.offline import init_notebook_mode,iplot
import plotly.graph_objects as pl
import cufflinks as cf
init_notebook_mode(connected=True)
```

```python
#labels
lab = df["gender"].value_counts().keys().tolist()
#values
val = df["gender"].value_counts().values.tolist()

trace = pl.Pie(labels=lab, values=val, 
               marker=dict(colors=['red']), 
                # Seting values to
               hoverinfo="value"
              )
data = [trace]
```

```python
layout = pl.Layout(title="Gender Distribution")
fig = pl.Figure(data = data,layout = layout)
iplot(fig)
```

```python
lab = df['test preparation course'].value_counts().keys().tolist()
val = df['test preparation course'].value_counts().values.tolist()
teace_test = pl.Pie(labels=lab, values=val, 
              marker=dict(colors=[ 'pink']),
              hoverinfo='value'
              )
data = [teace_test]
```

```python
layout = pl.Layout(title = 'test preparation distribution')
fig = pl.Figure(data=data, layout=layout)
iplot(fig)
```

```python
df_cross = pd.crosstab(df['race/ethnicity'],df["test preparation course"])

data = []

for x in df_cross.columns:
    data.append(pl.Bar(name=str(x), x=df_cross.index, y=df_cross[x]))

figure = pl.Figure(data)
figure.update_layout(barmode = 'group')
figure.show()
```
