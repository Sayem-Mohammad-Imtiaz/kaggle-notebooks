```python
df = pd.read_csv('/kaggle/input/processed-bbc-sport-dataset/bbcsport.csv')
df.head(2)
```

# Data preprocessing

```python
df['clean_text'] = hero.clean(df['text'])
```

# Custome cleaning

```python
from texthero import preprocessing as ppe
custom_pipeline = [ppe.fillna, ppe.lowercase, ppe.remove_whitespace]
df['clean_data'] = hero.clean(df['text'],custom_pipeline)
```

# TFIDF representation

```python
df['clean_text_tfidf'] = hero.do_tfidf(df['clean_text'])
```

# Principal Component Analysis

```python
df['clean_text_pca'] = hero.do_pca(df['clean_text_tfidf'])
```

# Scatter plot

```python
hero.scatterplot(df, col='clean_text_pca', color='topic', title="PCA BBC Sport news")
```
