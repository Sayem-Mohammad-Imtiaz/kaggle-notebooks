### Questão 1 - Item C - Representação Gráfica
Para cada uma das variáveis, produza um ou mais gráficos, usando matplotlib, que descreva seu comportamento / caracteristica. Lembre-se que estes gráficos precisam ser compatíveis com a classificação da variável. 

```python
import matplotlib.pyplot as plt
```

### 1) Variáveis usadas: aeronave_pmd_categoria e aeronave_tipo_veiculo

```python
g1 = df.groupby(['aeronave_tipo_veiculo'],as_index = False).sum()
g2 = df.groupby('aeronave_pmd_categoria',as_index = False).sum()
```

```python
# First Ring (outside)
fig, ax = plt.subplots()
ax.axis('equal')
mypie, _ = ax.pie(g2['total_fatalidades'], radius=1.9, labels=g2['aeronave_pmd_categoria'] )
plt.setp( mypie, width=0.3, edgecolor='white')
 
# Second Ring (Inside)
mypie2, _ = ax.pie(g1['total_fatalidades'], radius=1.9-0.3, labeldistance=0.7)
plt.setp( mypie2, width=0.4, edgecolor='white')
plt.margins(0,0)
plt.title('Fatalidades por Categoria/Tipo',loc='center',y=0.45)
plt.legend(mypie2,g1['aeronave_tipo_veiculo'],bbox_to_anchor=(1.2, 0.8),title='Tipos de Veículos')
# show it
plt.show()
```

### 2) Variável usada: aeronave_tipo_veiculo

```python
import numpy as np

g3 = df.groupby('aeronave_tipo_veiculo',as_index = False).sum().sort_values(by='total_fatalidades', ascending = False)

plt.rcdefaults()
fig, ax = plt.subplots()

# Data
y_pos = np.arange(len(g3['aeronave_tipo_veiculo']))
dados = g3['total_fatalidades']


ax.barh(y_pos, dados, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(g3['aeronave_tipo_veiculo'])
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('número de fatalidades')
ax.set_title('Fatalidades por tipo de veículo')
```

### 3) Variáveis usadas: aeronave_nivel_dano e aeronave_motor_quantidade

```python
g4 = df.groupby('aeronave_nivel_dano',as_index = False).sum().sort_values(by='total_fatalidades', ascending = False)
g5 = df.groupby('aeronave_motor_quantidade',as_index = False).sum().sort_values(by='total_fatalidades', ascending = False)


fig,axes = plt.subplots(nrows=1, ncols=2, figsize=(17, 6))


axes[0] = fig.add_axes([0,0,1,1])
axes[0].bar(g4['aeronave_nivel_dano'],g4['total_fatalidades'],color='b')
axes[0].set_ylabel('Fatalidades')

axes[1] = fig.add_axes([0,0,1,1])
axes[1].bar(g5['aeronave_motor_quantidade'],g5['total_fatalidades'],color='r')
axes[1].set_xlabel('Nível de Dano / Quantidade Motor')
plt.title('Fatalidade x Nível de dano e Quantidade Motor')
plt.show()
```

### 4) Variável usada: aeronave_operador_categoria

```python
g6 = df.groupby('aeronave_operador_categoria',as_index = False).sum().sort_values(by='total_fatalidades', ascending = False)

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = g6['aeronave_operador_categoria']
sizes = g6['total_fatalidades']
explode = (0.1, 0, 0, 0,0,0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, autopct='%1.1f%%', startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.legend(labels,bbox_to_anchor=(0.9, 0.8),title='Categorias de Operador:')
plt.title('Fatalidades por Categoria de Operador')
plt.show()
```

### 5) Variável usada: aeronave_fase_operacao

```python
g7 = df.groupby('aeronave_fase_operacao',as_index = False).sum().sort_values(by='total_fatalidades', ascending = False)
g8 = df.groupby('aeronave_fase_operacao',as_index = False).count().sort_values(by='total_fatalidades', ascending = False)[['aeronave_fase_operacao','total_fatalidades']]
g9 = g7.merge(g8,on='aeronave_fase_operacao',how='outer') #y são as ocorrencias
g9 = g9.sort_values(by='total_fatalidades_y',ascending= False)
g9 = g9[g9['total_fatalidades_y']>190]
```

```python
fig, ax1 = plt.subplots(figsize=(12,5))

#ax1.set_xlabel('Fase de Operação')
ax1.set_ylabel('Ocorrências', color='b')
ax1.bar(g9['aeronave_fase_operacao'],g9['total_fatalidades_y'])
ax1.tick_params(axis='y', labelcolor='b')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:red'
ax2.set_ylabel('Fatalidades', color=color)
ax2.plot(g9['aeronave_fase_operacao'],g9['total_fatalidades_x'],color='r')
ax2.tick_params(axis='y', labelcolor=color)
plt.title('Top 8: Nº de Ocorrências x Fatalidades por Fase de Operação')
fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

### 6) Variável usada: aeronave_pais_fabricante

```python
g10 = df.groupby('aeronave_pais_fabricante',as_index = False).sum().sort_values(by='total_fatalidades', ascending = False)
g10 = g10[g10['total_fatalidades']>16]
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(g10['aeronave_pais_fabricante'],g10['total_fatalidades'])
#ax.set_ylabel('Fatalidades')
plt.title('Top 5: Fatalidades por País Fabricante')
plt.show()
```
