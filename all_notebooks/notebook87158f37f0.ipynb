```python
data_enem = pd.read_csv("/kaggle/input/enem-2016/microdados_enem_2016_coma.csv", nrows=26000)
data_enem.head()
```

```python
list(data_enem.columns)
```

```python
new_data_enem = data_enem[["TP_SEXO", "NU_NOTA_REDACAO", "NU_IDADE"]] 
replace_dict = {'F':0,'M':1}
new_data_enem['TP_SEXO'][0] != 'M'
new_data_enem['TP_SEXO'] = new_data_enem['TP_SEXO'].map(replace_dict) 
form_data_enem = new_data_enem[~np.isnan(new_data_enem.NU_NOTA_REDACAO)]
form_data_enem["NU_NOTA_REDACAO"] = form_data_enem["NU_NOTA_REDACAO"].astype('int')
form_data_enem
```

```python
X = np.asarray(form_data_enem[['NU_NOTA_REDACAO', 'NU_IDADE']])
X[0:5]
```

```python
y = np.asarray(form_data_enem['TP_SEXO'])
y [0:5]
```

```python
from sklearn import preprocessing
X = preprocessing.StandardScaler().fit(X).transform(X)
X[0:5]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR
```

```python
yhat = LR.predict(X_test)
yhat
```

```python
yhat_prob = LR.predict_proba(X_test)
yhat_prob
```

```python
from sklearn.metrics import jaccard_score
jaccard_score(y_test, yhat,pos_label=0)
```
