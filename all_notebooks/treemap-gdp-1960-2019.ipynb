```python
GDP = pd.read_csv("/kaggle/input/gdp-annual-growth-for-each-country-1960-2020/GDP_annual_growth.csv")
```

```python
import pycountry # country locales for filtering non-countries

countries_code = [x.alpha_3 for x in pycountry.countries]
result = GDP['Country Code'].isin(countries_code)
n = []
for i in range(len(result)):
    if(result[i] == False):
        n.append(i)

GDP=GDP.drop(n, axis=0)
```

# Year 1960

```python
import plotly.express as px
import numpy as np

fig = px.treemap(GDP, path=['Country Name'], values='1960')
fig.show()
```
