```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('/kaggle/input/python-discord-pixels-analysis/pydis_pixels.csv')

counts = df['hex_code'].value_counts()  # Count for each hex code
counts  # Preview the data
```

```python
counts = counts[:15]  # Trim down to the top 15
counts.index = '#' + counts.index.astype(str)  # Add a # to each hex code for displaying colors

plt.xticks(rotation=55)  # Turn the labels sideways
plt.title('Pixels by color')  # Set a title

plt.bar(counts.index, counts.values, color=counts.index, edgecolor='black', linewidth=1)  # Graph the counts
plt.show()
```

```python
counts.drop(['#000000', '#ffffff'], inplace=True)  # Without black and white

plt.xticks(rotation=55)  # Turn the labels sideways
plt.title('Pixels by color without black or white')  # Set a title

plt.bar(counts.index, counts.values, color=counts.index, edgecolor='black', linewidth=1)  # Graph the counts
plt.show()
```
