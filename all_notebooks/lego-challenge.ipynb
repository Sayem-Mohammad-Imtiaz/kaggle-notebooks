### Construire un polynome et prédire

```python
x = parts_num().index
y = parts_num().num_parts
plt.scatter(x, y)
plt.show()
```

```python
z1 = np.polyfit(x,y,4)
p1 = np.poly1d(z1)
print(p1)
```

```python
yvals = p1(x)
plt.plot(x,y,'*',label='original values')
plt.plot(x,yvals,'red',label='polyfit values')
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.legend(loc = 4)
plt.show()
```
