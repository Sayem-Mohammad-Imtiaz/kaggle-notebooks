# Covid19 Vaccination Progress in Japan

```python
import numpy as np
import pandas as pd
import datetime as datetime
import matplotlib.dates as dates
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import os
```

```python
cases= pd.read_csv("../input/covid19-data-from-john-hopkins-university/CONVENIENT_global_confirmed_cases.csv")
cases[-3:]
```

```python
casesj = cases[['Country/Region','Japan']].copy()
casesj[-3:]
```

```python
casesj.columns=['date','cases']
casesj2 =casesj.drop([0])
casesj2['max cases']=casesj2['cases'].rolling(window=7).max()
casesj2['mean cases']=casesj2['cases'].rolling(window=7).mean()
casesj2['min cases']=casesj2['cases'].rolling(window=7).min()
casesj2[-3:]
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=casesj2['date'],y=casesj2['max cases'],name="7-day max"),secondary_y=False,)
fig.add_trace(go.Scatter(x=casesj2['date'],y=casesj2['mean cases'],name="7-day mean"),secondary_y=False,)
fig.add_trace(go.Scatter(x=casesj2['date'],y=casesj2['min cases'],name="7-day min"),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="Positive cases (7-day rolling) in Japan")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Cases",secondary_y=False)
fig.show()
```

```python
casesj2['date']=pd.to_datetime(casesj2['date'],format='%m/%d/%y') 
casesj2[-3:]
```

```python
vaccine= pd.read_csv("../input/covid-world-vaccination-progress/country_vaccinations.csv")
vaccinej = vaccine.loc[vaccine['country']=='Japan']
vaccinej[-3:]
```

```python
vaccinej2=vaccinej[['date','people_vaccinated','people_fully_vaccinated']].copy()
vaccinej2['people_vaccinated']=vaccinej2['people_vaccinated'].fillna(method='ffill')
vaccinej2['people_fully_vaccinated']=vaccinej2['people_fully_vaccinated'].fillna(method='ffill')
vaccinej2['people_average_vaccinated']=vaccinej2['people_vaccinated']
vaccinej2['people_average_fully_vaccinated']=vaccinej2['people_fully_vaccinated']
vaccinej2[-3:]
```

```python
population0= pd.read_csv("../input/population-by-country-2020/population_by_country_2020.csv")
population = population0.loc[population0['Country (or dependency)']=='Japan']
population
```

```python
pop0= population['Population (2020)'].values[0]
print(pop0)
```

```python
vaccinej2['% people_average_vaccinated'] = vaccinej2['people_average_vaccinated']*100/pop0
vaccinej2['% people_average_fully_vaccinated'] = vaccinej2['people_average_fully_vaccinated']*100/pop0
vaccinej2[-3:]
```

```python
fig = make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=vaccinej2['date'],y=vaccinej2['people_average_vaccinated'],name='People vaccinated' ),secondary_y=False,)
fig.add_trace(go.Scatter(x=vaccinej2['date'],y=vaccinej2['people_average_fully_vaccinated'],name='People fully vaccinated '),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="Vaccination Progress in Japan")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Vaccinated",secondary_y=False)
fig.show()
```

```python
fig = make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=vaccinej2['date'], y=vaccinej2['% people_average_vaccinated'],name ='% People vaccinated' ),secondary_y=False,)
fig.add_trace(go.Scatter(x=vaccinej2['date'], y=vaccinej2['% people_average_fully_vaccinated'],name='% People fully vaccinated'),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="% Vaccination Progress in Japan")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="% Vaccinated",secondary_y=False)
fig.show()
```

* A mysterious jump was observed on 2021-08-04 in % People vaccinated.

* The population aged 65 and over is 36.17 million, which is 28.7% of the population. 
* As of mid-May in 2021, the Japanese government declared that vaccination for them would be completed by the end of July. If it was done as scheduled,% people_average_fully_vaccinated would reach 30% by that time. ==> At the end of July, the ratio reached 29.9%. Pass!!! 

# Vaccination Progress World Ranking

```python
vaccine3=vaccine[['country','people_fully_vaccinated_per_hundred','date']].copy()
vaccine4=vaccine3.groupby(['country'], as_index=False).max()
vaccine5=vaccine4.sort_values('people_fully_vaccinated_per_hundred',ascending=False)
rank=vaccine5[vaccine5['country']=='Japan'].index.values[0]+1
date=vaccine5[vaccine5['country']=='Japan']['date'].values[0]
print('Vaccination progress of Japan is the '+str(rank)+'th in the world rankings as of '+str(date)+'.' )
```
