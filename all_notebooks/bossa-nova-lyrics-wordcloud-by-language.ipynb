```python
import pandas as pd
from pandas.io.json import json_normalize
import warnings
warnings.filterwarnings("ignore")
from sklearn.feature_extraction.text import CountVectorizer
import random
from wordcloud import WordCloud 
import matplotlib.pyplot as plt
import seaborn as sns
import json_lines
from tqdm.notebook import tqdm
```

```python
data=pd.read_csv('../input/bossa-nova-lyrics/bossa_nova_songs_dataset.csv')
data
```

```python
langs=list(data['song_lang'].unique())
print(langs)
```

```python
data['song_lyrics'][0]
```

```python
import re
import nltk
nltk.download("stopwords")
nltk.download('punkt')
from nltk import word_tokenize,sent_tokenize
nltk.download('wordnet')
import nltk as nlp
```

```python
cult_list=[]
for item in langs[0:4]:
    df=data['song_lyrics'][data['song_lang']==item]
    cult_list0=[]
    for cult in tqdm(df):
        cult=re.sub("[^a-zA-z]"," ",str(cult))
        cult=cult.lower()
        cult=nltk.word_tokenize(cult)
        lemma=nlp.WordNetLemmatizer()
        cult=[lemma.lemmatize(word) for word in cult]
        cult=" ".join(cult)
        cult_list0.append(cult)
    cult_list+=[cult_list0]
```

```python
max_features=400
count_vectorizer=CountVectorizer(max_features=max_features,stop_words="english")
data4=[]
for i in range(4):
    sparce_matrix=count_vectorizer.fit_transform(cult_list[i]).toarray()
    data4+=[pd.DataFrame(count_vectorizer.get_feature_names(),columns=["Words"])]
```

```python
fig,axs = plt.subplots(2,2,figsize=(16,12))
for i in range(4):
    r=i//2
    c=i%2
    random.seed(2021)
    random.shuffle(data4[i].Words)
    wordcloud=WordCloud(background_color="black",width=1024,height=768).generate(" ".join(data4[i].Words))    
    ax=axs[r][c].axis("off")  
    ax=axs[r][c].set_title(langs[i])
    ax=axs[r][c].imshow(wordcloud)
plt.show()
```
