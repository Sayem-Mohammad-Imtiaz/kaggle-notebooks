```python
import numpy as np
from keras.datasets import mnist
```

```python
(xtrain, ytrain), (xtest, ytest) = mnist.load_data()
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
xtrain
```

```python
ytrain
```

```python
ytrain = ytrain.reshape(-1,1)
ytest = ytest.reshape(-1,1)
```

```python
from keras import models
from keras import layers
```

We have 2 different dense layers (fully connected layers) and they follow each other. Second layer (last layer) has 10 different outputs and if we sum them up, they give us 1  as a probabilistic value.   

According to this information, I am going to create my model:

```python
n_network = models.Sequential()
```

```python
n_network
```

```python
n_network.add(layers.Dense(512, activation = "relu", input_shape = (28*28,)))
n_network.add(layers.Dense(10, activation = "softmax"))
```

```python
n_network.compile(optimizer = "rmsprop", loss = "categorical_crossentropy", metrics = ["accuracy"])
```

```python
xtrain.shape
```

I am going to convert unit8 (ranged as [0,255]) pixel values to float32 (ranged as [0,1]) data type.

```python
xtrain = xtrain.reshape((60000, 28*28))
```

```python
xtrain
```

```python
xtrain = xtrain.astype("float32")/255
```

```python
xtest.shape
```

```python
xtest.dtype
```

```python
xtest = xtest.reshape((10000, 28*28))
```

```python
xtest.shape
```

```python
xtest = xtest.astype("float32") / 255
```

```python
xtest.dtype
```

Besides, our ytrain and ytest values should be categorical values:

```python
ytrain
```

```python
from keras.utils import to_categorical
```

```python
ytrain = to_categorical(ytrain)
```

```python
ytrain
```

```python
ytest = to_categorical(ytest)
```

```python
ytest
```

```python
n_network.fit(xtrain, ytrain, epochs = 10, batch_size = 128)
```

Let see the results for test set:

```python
n_network.evaluate(xtest, ytest)
```
