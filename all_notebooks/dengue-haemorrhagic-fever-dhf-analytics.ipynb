```python
data = pd.read_csv('../input/patient-diagnosis-at-one-of-hospital-in-bali/patient diagnosis.csv')
data.head()
```

****Get icd-10 code A91 Dengue haemorrhagic fever (DHF)

```python
dhf=data[data.Icd_10=='A91']
dhf
```

****Extract Year

```python
from datetime import datetime
pd.options.mode.chained_assignment = None  # default='warn'
year_in=dhf["Date_in"]
years = [datetime.strptime(date, '%Y-%m-%d').year for date in year_in]
dhf["Year"]=years
dhf
```

```python
year_list = dhf['Year'].unique()
year_list
```

****count the diagnosis data group by year

```python
diagnosis_number=dhf.groupby('Year').count().Diagnosis
diagnosis_number
```

****Line chart result for DHF diagnosis every year

```python
plt.plot(year_list,diagnosis_number)
plt.show()
```

> 2015 and 2016 is the most of DHF diagnosis in this Hospital
