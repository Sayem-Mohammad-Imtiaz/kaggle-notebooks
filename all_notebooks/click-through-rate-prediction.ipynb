```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
x=df.drop(['Ad Topic Line','City','Timestamp','Country','Clicked on Ad'],axis=1)
y=df['Clicked on Ad']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
l = LogisticRegression()
l.fit(x_train,y_train)
```

```python
pred = l.predict(x_test)
```

```python
pred
```

```python
df1=pd.DataFrame({'Actual Value':y_test,'Predicted Value':pred})
df1.head()
```

```python
v=[]
for x in pred:
    if x==1:
        v.append('Yes,Clicked!')
    else:
        v.append('Not Clicked!')

df1['Clicked Or Not']=v
df1.head(10)
```

```python
acc_logreg = round(accuracy_score(pred, y_test)*100, 2)
acc_logreg
```

```python
print(classification_report(y_test, pred))
print(confusion_matrix(y_test, pred))
```
