```python
#This notebook deals with analyzing various parameters which affect Graduate Admissions.
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#The chance of getting admitted into a university depends on the following parameters:
#1.GRE SCORE
#2.TOEFL SCORE
#3.University Ranking
#4.SOP
#5.LOR
#6.CGPA
#7.Research
#8.Chance of Admit

#Let's now analyze how each of these parametes affets admissions
```

```python
data = pd.read_csv("../input/us-graduate-schools-admission-parameters/US_graduate_schools_admission_parameters_dataset.csv")
data.head()
```

```python
#Let's start with exploring the dataset
data.shape
```

```python
data.describe()
```

```python
#Now let's see how GRE score affects admission
print(data.columns)
import seaborn as sns
sns.distplot(a = data['GRE Score'] , kde = False)
```

```python
#An ideal GRE score of around 325-330 has the maximum number of admits into universities.
```

```python
#Analysis of SOP vs Chance of Admit
sns.barplot(x=data['SOP'],y=data['Chance of Admit '])
```

```python
#A statement of purpose is a must for getting into a good university. The above plot shows that the chance of admit increases dramatically when the quality of SOP is better.
```

```python
#Now let's see the influence of Research on Chance of Admit
sns.swarmplot(x = data['Research'],y = data['Chance of Admit '])
```

```python
#The swarm plot shows that a solid research profile is necessary to get an admission offer from a top university.
```

```python
sns.heatmap(data = data)
```

```python
#This heatmap makes it evident that a TOEFL score of around 108 has the maximum probability of securing an admission
```

```python
sns.barplot(x=data['University Rating'],y=data['Chance of Admit '])
```

```python
#This simple barplot visualization shows the relationship between university rating and chance of admit. As the university rating increases the chance of admit also increases marginally
```

```python
sns.barplot(x=data['Research'],y=data['Chance of Admit '])
```

```python
#This barplot proves that research experience plays a very important role in graduate admissions and is one of the most important parameter.
```

```python
sns.kdeplot(data = data['LOR '],shade = True)
```

```python
#A strong letter of recommendation is always a plus point in securing an admit. This plot proves it here.
```

```python
sns.distplot(a = data['CGPA'] , kde = False)
```
