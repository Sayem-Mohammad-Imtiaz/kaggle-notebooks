```python
import pandas as pd

#origin of dataset: https://data.ny.gov/Transportation/Vehicle-Snowmobile-and-Boat-Registrations/w4pv-hbkt

data = pd.read_csv("../input/ny-vehicles-registrations/Vehicle__Snowmobile__and_Boat_Registrations.csv")
modern_cars = data[data.VIN.str.len()==17]
```

```python
e39m5 = modern_cars[modern_cars.VIN.str.startswith("WBSDE93")]
```

```python
e39m5.groupby(["Model Year"]).count()["VIN"]
```

```python
e39m5.groupby(["Color"]).count()["VIN"]
```
