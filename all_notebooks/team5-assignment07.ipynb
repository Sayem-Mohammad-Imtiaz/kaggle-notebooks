```python
#1

import pyspark

from pyspark.sql import SparkSession
from pyspark.sql import functions as f

spark = SparkSession.builder.appName("Team5").getOrCreate()
movies = (spark.read.csv(path='/kaggle/input/msis405l21cttt-spring-2021-assignment-7/movies_small.csv',
                           sep=',', header=True, quote='"', schema="movieId INT, title STRING, genres STRING" ))
```

```python
movies.filter(f.array_contains(f.split(f.col('genres'), '\|'), 'Action')).show(5, False)
```

```python
#2
movies.withColumn('count', f.size(f.split(f.col('genres'), '\|'))).select('movieId', 'title', 'count').show(5, False)
```

```python
#3
movies.withColumn("genre", f.explode(f.split(f.col("genres"), '\|'))) \
.groupBy('genre').agg(f.count('title').alias('count')).sort('count').show(5, False)
```

```python
#4
movies.withColumn("genre", f.explode(f.split(f.col("genres"), '\|'))).groupBy('genre').\
agg(f.collect_list('title').alias('movie_list')).show(truncate=100)
```

```python
movies.withColumn("genre", f.explode(f.split(f.col("genres"), '\|'))).\
withColumn('title', f.split(f.col('title'), ' ')).withColumn('title', f.element_at('title', -1)).\
withColumn('first_appearance', f.substring('title', 2, 4)).filter(f.col('genre').isin(['Animation', 'Sci-Fi'])).\
groupBy('genre').agg(f.min(f.col('first_appearance').cast('int')).alias('first_appearance')).show()
```
