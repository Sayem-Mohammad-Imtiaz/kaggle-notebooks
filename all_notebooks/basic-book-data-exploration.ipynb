# Loading in the data and removing rows with errors

```python
bkdf = pd.read_csv('/kaggle/input/books.csv',error_bad_lines = False)
```

```python
bkdf.info()
```

# Determining the number of unique average ratings 

```python

x = len(list(set(bkdf['average_rating'])))
print(x)
```

# Reindexing the DF and removing the book ID column and then grouping by average rating

```python
bkdf.reindex(bkdf.iloc[:,0])
bkdf.drop('bookID',axis=1,inplace=True)
(bkdf.groupby('average_rating').count()).loc[:,'ratings_count']
```

# Plotting the average rating data as a histogram

```python
bkdf.plot(y = 'average_rating', kind = 'hist',bins = 50)
plt.xlabel('Average Rating')
plt.ylabel('Count')
```

# Looking at the correllation between two variables 

```python
bkdf.plot(x = 'average_rating',y='ratings_count', kind = 'scatter')
```

# Plotting a box plot of the average ratings

```python
bkdf.plot(y = 'average_rating', kind = 'box')
```

```python
bkdf2 = bkdf[bkdf['average_rating'] == 5]
bkdf2.head()
```
