```python
# ein einfacher Graph
from random import random

x = np.arange(10)
y = np.array([2*i + 3*random() for i in x])

plt.plot(x,y)
plt.show()
```

```python
# Scatter- plot
x = np.arange(10)
y = np.array([2*i + 3*random() for i in x])

plt.scatter(x,y)
plt.show()
```

```python
females = list(data.columns[2:10])
males = list(data.columns[11:19])

males
```

```python
labels = list(data.columns[20:])

for x in labels:
    labels[labels.index(x)] = x.replace("person", "", 1)
    x = x.replace("person", "", 1)
    labels[labels.index(x)] = x.replace("and", "\n")
    x = x.replace("and", "\n")
    labels[labels.index(x)] = x.title()
    
print(labels)
```

```python
diff = [np.array(data[m]) - np.array(data[f]) for f, m in zip(females, males)]
summe = [np.array(data[m]) + np.array(data[f]) for f, m in zip(females, males)]

diff = np.array(diff)
summe = np.array(summe)
```
