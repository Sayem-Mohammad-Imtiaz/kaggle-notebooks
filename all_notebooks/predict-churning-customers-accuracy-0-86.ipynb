```python
df=pd.read_csv('../input/credit-card-customers/BankChurners.csv')
```

```python
df.head()
```

```python
df=df.drop('Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2',axis=1)
```

```python
df=df.drop('Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',axis=1)
```

```python
plt.plot(df.isnull())
```

```python
sns.countplot(x='Income_Category',data=df)
```

```python
from sklearn.preprocessing import LabelBinarizer

encoder = LabelBinarizer()
encoder.fit(df['Income_Category'])
transformed = encoder.transform(df['Income_Category'])
ohe_df = pd.DataFrame(transformed)
df = pd.concat([df, ohe_df], axis=1).drop(['Income_Category'], axis=1)
```

```python
education=pd.get_dummies(df['Education_Level'],drop_first=True)
df=pd.concat([df,education],axis=1)
```

```python
df.head()
df=df.drop('Education_Level',axis=1)
```

```python
card=pd.get_dummies(df['Card_Category'],drop_first=True)
df=pd.concat([df,card],axis=1)
df=df.drop(['Card_Category','Marital_Status'],axis=1)
```

```python
df=df.drop('Gender',axis=1)
```

```python
Churn=pd.get_dummies(df['Attrition_Flag'],drop_first=True)
df=pd.concat([df,Churn],axis=1)
df=df.drop('Attrition_Flag',axis=1)
```

standarize the Total_Revolving_Bal

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df[['Total_Revolving_Bal']])
scaled_df=scaler.transform(df[['Total_Revolving_Bal']])
scaled_df=pd.DataFrame(scaled_df)
df=pd.concat([df,scaled_df],axis=1)
```

```python
df=df.drop('Total_Revolving_Bal',axis=1)
```

```python
x=df.drop(['CLIENTNUM','Existing Customer'],axis=1)
```

```python
y=df['Existing Customer']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
predictions=logmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```
