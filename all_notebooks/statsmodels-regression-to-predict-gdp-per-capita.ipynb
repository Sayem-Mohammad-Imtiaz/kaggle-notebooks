```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Creating a regression model to predict GDP per capita ##

```python
dados = pd.read_csv("../input/countries-of-the-world/countries of the world.csv", decimal = ',')
```

```python
dados.head()
```

### Tranforming data, dropping NAs and categorical data ###

```python
encoder = LabelEncoder()

dados_transformados = dados
dados_transformados = dados_transformados.fillna(0)
dados_transformados = dados_transformados.drop(['Country',"Region"], 1)
dados_transformados.head()
```

### Plotting heatmap to choose varibles ###

```python
ax, fig = plt.subplots(figsize =(18,9))
sns.heatmap(dados_transformados.corr(), annot = True)
plt.title("Correlation HeatMap")
```

```python
import statsmodels.formula.api as sm
from statsmodels.regression import linear_model
import statsmodels.api as api
```

### Creating endog and exog variables ###

```python
x = dados_transformados[["Phones (per 1000)", "Net migration", "Agriculture"]]
y = dados_transformados["GDP ($ per capita)"]
x = api.add_constant(x)
```

```python
z = dados.loc[:,["Region", "Phones (per 1000)", "Net migration", "Agriculture", "GDP ($ per capita)"]]
sns.pairplot(z, hue="Region", palette="Paired")
```

```python
modelo = linear_model.OLS(endog=y, exog=x).fit()
```

```python
print(modelo.summary())
```
