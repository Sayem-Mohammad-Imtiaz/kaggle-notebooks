****    Data Manuiplation****
Change all Headers into upper case and as we can in date column time is constant throught the entire dataset so we can remove time part from date

```python
df_border_cross.columns=map(str.upper,df_border_cross.columns)
df_border_cross.head()
```

```python
df_border_cross['DATE']=pd.to_datetime(df_border_cross['DATE'])
df_border_cross['YEAR']=df_border_cross['DATE'].dt.year
df_border_cross['MONTH']=df_border_cross['DATE'].dt.month
df_border_cross['DAY']=df_border_cross['DATE'].dt.day
df_border_cross.head()
```

```python
rm_duplicated_data=df_border_cross.drop_duplicates()
rm_duplicated_data.count()
```

```python
temp=pd.DataFrame(df_border_cross.groupby('BORDER')['VALUE'].sum().sort_values()).reset_index()
sns.barplot(x = 'BORDER', y = 'VALUE', data = df_border_cross)
```

```python
ax=sns.lineplot(x = 'YEAR' , y = 'VALUE', hue = 'MEASURE', data = df_border_cross, legend='full')
ax.legend(loc='best', bbox_to_anchor=(1.05, 1),
          ncol=2, fancybox=True, shadow=True)
```

```python
sns.lineplot(x='MONTH',y='VALUE',data=df_border_cross)
```

```python
temp = pd.DataFrame(df_border_cross.groupby('MEASURE')['VALUE'].sum().sort_values(ascending = False)).reset_index()
fig = plt.figure(figsize = (10,5))
ax = sns.barplot(x = 'MEASURE', y = 'VALUE', data = temp)
var = ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
```

```python
temp_v2=pd.DataFrame(df_border_cross.groupby('PORT NAME')['VALUE'].sum().sort_values(ascending=False)).reset_index()
fig = plt.figure(figsize = (25,15))
ax_v2=sns.barplot(x='PORT NAME',y='VALUE',data=temp_v2)
var=ax_v2.set_xticklabels(ax_v2.get_xticklabels(), rotation = 90)
```
