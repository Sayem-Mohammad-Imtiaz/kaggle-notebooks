# Exploratory Data Analysis
#### **What is the overall sales trend**

### **1.How many cars were sold in Canada in the last 3 years ?**

```python
# How many car was sold in Canada in the last 3 years
df.groupby('Year').sum()['Sumofsales'].reset_index
```

### **2. Most sold and least sold car**

```python
# The Maximum sold car 
df.nlargest(1, ['Sumofsales'])
#Max is Jeep Wrangler in 2019
```

```python
# The Minimum sold car
df.nsmallest(1, ['Sumofsales'])
#Min is Volkswagen Touareg in 2020
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(df['Year'],df['Sumofsales'])
plt.show()
```

### **3. What are the top 10 Cars by sales ?**

```python
# Top 10 Cars by sales
df.nlargest(10, ['Sumofsales'])
```

### **4. What is the most prefable category ?**

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(df['Category'],df['Sumofsales'])
plt.show()
## MIDSIZE SUV is the most prefable category
```
