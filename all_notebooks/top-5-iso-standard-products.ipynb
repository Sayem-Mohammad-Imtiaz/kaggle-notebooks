Calculate `price_CF` to `Number of pages` ratio and its mean

```python
# Filter off rows with na prices
data = data.loc[data['price_CF'].notna()]
data.isna().sum()
```

```python
data['price_to_pages_ratio'] = data['price_CF'] / data['Number of pages']
price_to_pages_ratio_mean = data['price_to_pages_ratio'].mean()
price_to_pages_ratio_mean
```

Filter for Published `Status` with `price_CF` to `Number of pages` ratio greater than its mean

```python
data = data.loc[pd.DataFrame([data['Status'] == 'Published', 
                              data['price_to_pages_ratio'] > price_to_pages_ratio_mean])
                .all()]
data.head()
```
