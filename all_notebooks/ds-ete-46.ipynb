```python
#importing the dataset
dataset = pd.read_csv('../input/suicide-rates-overview-1985-to-2016/master.csv')
dataset.head()
```

```python
#describing the data
dataset.describe()
```

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-whitegrid") 
import seaborn as sns
```

```python
dataset.shape
```

```python
dataset.size
```

```python
dataset.columns
```

```python
dataset.rename(columns={'suicides_no':'suicides_number',
                            'suicides/100k pop':'suicides_per_100k_pop',
                            ' gdp_for_year ($) ':'gdp_for_year', 
                       'gdp_per_capita ($)':'gdp_per_capital'}, inplace=True)
dataset.columns = dataset.columns.str.replace(" ","_")
dataset.columns = dataset.columns.str.lower()
dataset.columns
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
#mean of sucides in males and females
dataset.groupby("sex")["suicides_number"].mean()
```

```python
#suicides per 100k population in males and females
dataset.groupby("sex")["suicides_per_100k_pop"].mean()
```

```python
#graph of the suicides of male and female
plt.figure(figsize=(8,5))
sns.barplot(x = "sex", 
            y = "suicides_number", 
            data = dataset)
plt.xlabel('Gender')
plt.ylabel('Number of Suicide')
plt.title('Suicide Numbers by Gender')
plt.show()
```

```python
#suicides numbers in each year
dataset.groupby("year", as_index = False)["suicides_number"].mean().sort_values(by="suicides_number",ascending = False)
```

```python
yearMeanSN = dataset.groupby("year", as_index = False)["suicides_number"].mean()
plt.figure(figsize=(15,6))
sns.pointplot(x = "year", 
              y = "suicides_number", 
              data = yearMeanSN)
plt.xlabel('Year')
plt.ylabel('Average Number of Suicide')
plt.title('Average Suicide Numbers by Years')
plt.show()
```

```python

dataset.groupby("age", as_index = False)["suicides_per_100k_pop"].mean().sort_values(by="suicides_per_100k_pop",ascending = False)
```

```python
#suicides in age group
ageDataSN = dataset.groupby("age", as_index = False)["suicides_number"].mean().sort_values(by="suicides_number",ascending = False)
plt.figure(figsize=(8,5))
sns.barplot(x = "age", 
            y = "suicides_number", 
            data = ageDataSN)
plt.xlabel('Age Group')
plt.ylabel('Number of Suicide')
plt.title('Suicide Numbers by Age Groups')
plt.show()
```

```python
dataset["generation"].value_counts()
```

```python
dataset.groupby("generation", as_index = False)["suicides_number"].mean().sort_values(by="suicides_number",ascending = False)
```

```python
genDataSN = dataset.groupby("generation", as_index = False)["suicides_number"].mean().sort_values(by="suicides_number",ascending = False)
plt.figure(figsize=(8,5))
sns.barplot(x = "generation",
            y = "suicides_number",
            data = genDataSN)
plt.xlabel('Generation')
plt.ylabel('Number of Suicide')
plt.title('Suicide Numbers by Generations')
plt.show()
```
