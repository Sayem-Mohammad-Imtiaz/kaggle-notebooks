```python
#Importando as bibliotecas
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
# Carregando o dataframe
diamonds = pd.read_csv("../input/diamonds/diamonds.csv")
```

```python
#Listando os dados
diamonds.head(5)
```

```python
#Renomeando as colunas
diamonds.rename(columns={'color': 'Cor'}, inplace = True)
diamonds.rename(columns={'clarity': 'Clareza'}, inplace = True)
diamonds.rename(columns={'cut': 'Corte'}, inplace = True)
diamonds.rename(columns={'price': 'Preço'}, inplace = True)
diamonds.rename(columns={'carat': 'Quilate'}, inplace = True)
diamonds.rename(columns={'depth': 'Profundidade'}, inplace = True)
```

```python
# removendo as colunas id, gênero e idade
var = diamonds.drop(['Unnamed: 0'], axis = 1)
var
```

```python
diamonds.info()
```

```python
#figsize representa a largua x altura
plt.figure(figsize=[14,10])


plt.subplot(221)
plt.hist(var['Quilate'],bins=20,color='gray')
plt.xlabel('Quilate')
plt.ylabel('Frequência')
plt.title('Frequência de diamantes por quilate')


plt.subplot(222)
plt.hist(var['Profundidade'],bins=20,color='blue')
plt.xlabel('Profundidade')
plt.ylabel('Fequência')
plt.title('Frequência de diamantes por profundidade')


plt.subplot(223)
plt.hist(var['Preço'],bins=20,color='yellow')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.title('Frequência de diamantes por Price')


plt.subplot(224)
plt.hist(var['Corte'],bins=20,color='green')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.title('Frequência de diamantes por Corte')
```

```python
df_regressao = pd.get_dummies(diamonds, columns=["Corte", "Cor", "Clareza"])
```

```python
df_regressao
```

```python
X = df_regressao.drop(['Preço'],1)
y = df_regressao['Preço']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4)
```

```python
model = LinearRegression()
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
