## Replacing all the NAN values with 0. 

The Dataset consists of all numbers. Any missing data can be marked as 0 (with the assumption that there were no vaccinations provided on that particular day).

```python
df.loc[np.isnan(df['total_vaccinations']),['total_vaccinations']] = 0
df.loc[np.isnan(df['people_vaccinated']),['people_vaccinated']] = 0
df.loc[np.isnan(df['people_fully_vaccinated']),['people_fully_vaccinated']] = 0
df.loc[np.isnan(df['daily_vaccinations_raw']),['daily_vaccinations_raw']] = 0
df.loc[np.isnan(df['daily_vaccinations']),['daily_vaccinations']] = 0
df.loc[np.isnan(df['total_vaccinations_per_hundred']),['total_vaccinations_per_hundred']] = 0
df.loc[np.isnan(df['people_vaccinated_per_hundred']),['people_vaccinated_per_hundred']] = 0
df.loc[np.isnan(df['people_fully_vaccinated_per_hundred']),['people_fully_vaccinated_per_hundred']] = 0
df.loc[np.isnan(df['daily_vaccinations_per_million']),['daily_vaccinations_per_million']] = 0
```

## Calculating the Sum of all Columns for Each Country

```python
df.groupby([df.country]).sum().sort_values(by='total_vaccinations',ascending=False)
```

## Top 5 Countries for Daily Vaccinations

```python
plt.figure(figsize=[10,10])
df_top5 = df.groupby([df.country]).sum().sort_values(by='total_vaccinations',ascending=False)[0:10]
df_top5
df_top5['daily_vaccinations'].plot.bar()
plt.show()
```

## Countries at the end of the Covid19 Vaccionation drive

```python
plt.figure(figsize=[10,10])
df_top5 = df.groupby([df.country]).sum().sort_values(by='total_vaccinations',ascending=False)[-11:-1]
df_top5
df_top5['daily_vaccinations'].plot.bar()
plt.show()
```
