# Data preprocessing

```python
#Replacing whitespaces with nan
df=df.replace(r'^\s*$', np.nan, regex=True)
```

```python
# Checking if there any null values
df.isnull().sum()
```

* TotalCharges has 11 null values

```python
#Converting TotalCharges column to integer type
df['TotalCharges'] = pd.to_numeric(df['TotalCharges']) 
```

```python
#Dropping the null values
df = df.dropna()
df.isnull().sum()
```

```python
#Checking if there are any duplicate records
df[df.duplicated()]
```
