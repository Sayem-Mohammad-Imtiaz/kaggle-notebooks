R^2 ifadesi: bağımsız değişkenlerin bağımlı değişkendeki değişikliği açıklama yüzdesidir. 0.977 oldukça yüksek bir rakam.
satışlardaki değişimi çok iyi açıklayabiliriz demek oluyor.
R^2 ne kadar değişken eklenirse o kadar şişmektedir. Şişmenin önüne geçmek için düzeltilmiş R^2 değeri ortaya çıkar. Daha sağlıklı bir açıklanabilirlik oranı vermektedir. 
F-statistic: modelin anlamlılığını test etmek için kullanılan istatistiktir. 
Prob(F- statistic: 0.05 den oldukça küçük bir değere sahip. Bu yüzden modelin anlamlı olduğu bilgisini bize veriyor. (modelin anlamlılığıyla ilgili bize bilgi veriyor.)


coef: bağımsız değişkenlerin katsayılarını ifade ediyor. (b1, b2, b3 katsayılarını ifade ediyor.) 
std err: bağımsız değişkenlerin standart hata değeridir.
t istatistiği anlamlılığı ifade eder.
P>t modellemek için kullandığımız bütün değişkenler anlamlıdır. 
0.05 ten küçük değerler elde ettiğimiz sürece anlamlı değişkenler olur.
**coef katsayısı yorumu:** örn: Tv reklamını 1 birimlik yatırım yapıldığı zaman 0.0671 bir artış söz konusu olacaktır.

**MODEL TUNING**(model doğrulama)

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.20, random_state =99)
```

```python
X_train.head() #train setitin bağımsız değişkenleri
```

```python
y_train.head()
```

```python
X_test.head()
```

```python
y_test.head()
```

```python
lm= LinearRegression()
model= lm.fit(X_train, y_train)
```

```python
#train(eğitim) hatası 
np.sqrt(mean_squared_error(y_train, model.predict(X_train)))
```

```python
#test hatası 
np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(model, X_train, y_train, cv=10, scoring="neg_mean_squared_error")
```

```python
np.mean(-cross_val_score(model, X_train, y_train, cv=10, scoring="neg_mean_squared_error"))
#ortalama hata
```
