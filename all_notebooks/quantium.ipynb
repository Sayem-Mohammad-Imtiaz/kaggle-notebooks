```python
def age_cat(a):
    b=a.split(" ")
    Age=''
    if b[0]=='NEW':
        Age='MIDAGE';
    elif b[0]=='RETIREES':
        Age='OLDER'
    else:
        Age=b[0]
    return Age;

def Relation_cat(a):
    b=a.split(" ")
    Status=''
    if b[-1]=='SINGLES/COUPLES':
         Status='SINGLES/COUPLES';
    elif b[0]=='RETIREES':
         Status='RETIREES'
    else:
         Status='FAMILIES'
    return Status
```

```python
def get_month(a):
    b=a.month
    return b

def Big(st_num):
    st_Data=Data[Data['STORE_NBR']==st_num]
    st_Data['month']=st_Data['Dates'].apply(get_month)
    
    st_Data_total=st_Data.groupby(['month']).sum()
    st_Data_total = st_Data_total.rename(columns = {'TOT_SALES': 'Total Sales'}, inplace = False)
    st_Data_total=st_Data_total.drop(['LYLTY_CARD_NBR','TXN_ID','PROD_NBR','STORE_NBR','PROD_QTY'],axis=1)
    
    st_Data_cust=pd.DataFrame(st_Data.groupby('month').LYLTY_CARD_NBR.nunique())
    st_Data_cust =st_Data_cust.rename(columns = {'LYLTY_CARD_NBR': 'Total Customers'}, inplace = False)
    #st_Data_cust=st_Data_cust.drop(['LYLTY_CARD_NBR','TXN_ID','PROD_NBR','STORE_NBR','PROD_QTY'],axis=1)
    
    st_Data_tra=st_Data.groupby(['month']).count()
    st_Data_tra= st_Data_tra.rename(columns = {'TXN_ID': 'Total Transactions'}, inplace = False)
    st_Data_tra=st_Data_tra.drop(['LYLTY_CARD_NBR', 'PROD_NBR', 'PROD_NAME', 'PROD_QTY',
       'TOT_SALES', 'Brand Name', 'Packet Size', 'Dates', 'LIFESTAGE',
       'PREMIUM_CUSTOMER', 'Age Category', 'Type','STORE_NBR'],axis=1)
    st_Data_tra['Transaction per Customer']=st_Data_tra['Total Transactions']/st_Data_cust['Total Customers']
    st_Data_tra=st_Data_tra.drop(['Total Transactions'],axis=1)
    
    #sns.lineplot(x=st_Data_total['month'],y=st_Data_total['Total Sales'])
    #st_Data_total.plot(figsize=(15,3))
    #st_Data_cust.plot(figsize=(15,3))
    #st_Data_tra.plot(figsize=(15,3))
    
    st_dfs = [st_Data_total, st_Data_cust, st_Data_tra]
    st_df_final = reduce(lambda left,right: pd.merge(left,right,on='month'), st_dfs)
    return st_df_final
```

```python

       
def cor_st_sale(num):
    total_corr_sale=[]
    for i in include_list:
        r = np.corrcoef(Big(num)['Total Sales'], Big(i)['Total Sales'])
        total_corr_sale.append(r[0, 1])
    total_corr_sale_frame=pd.DataFrame(total_corr_sale,columns=['Sale corr'])
    include_list_frame=pd.DataFrame(include_list,columns=['Store Number'])
    df_sale_corr=pd.concat([total_corr_sale_frame,include_list_frame],axis=1)
    #total_corr_sale_frame,include_list_frame
    return df_sale_corr
        

def cor_st_cust(num):
    total_corr_cust=[]
    for i in include_list:
        r = np.corrcoef(Big(num)['Total Customers'], Big(i)['Total Customers'])
        total_corr_cust.append(r[0, 1])
    total_corr_cust_frame=pd.DataFrame(total_corr_cust,columns=['Customer corr'])
    include_list_frame=pd.DataFrame(include_list,columns=['Store Number'])
    df_cust_corr=pd.concat([total_corr_cust_frame,include_list_frame],axis=1)
    return df_cust_corr
    

def cor_st_trans(num):
    total_corr_trans=[]
    for i in include_list:
        r = np.corrcoef(Big(num)['Transaction per Customer'], Big(i)['Transaction per Customer'])
        total_corr_trans.append(r[0, 1])
    total_corr_trans_frame=pd.DataFrame(total_corr_trans,columns=['Tran/cust corr'])
    include_list_frame=pd.DataFrame(include_list,columns=['Store Number'])
    df_trans_corr=pd.concat([total_corr_trans_frame,include_list_frame],axis=1)
    return df_trans_corr
        
def corr_st(st_num):
    df1=cor_st_cust(st_num)
    df2=cor_st_sale(st_num)
    df3=cor_st_trans(st_num)
    dfs = [df1,df2,df3]
    corr_df = reduce(lambda left,right: pd.merge(left,right,on='Store Number'), dfs)
    return corr_df
 
    
```
