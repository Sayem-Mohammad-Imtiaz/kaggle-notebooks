# 2. Data Visualization

Due to all the data are numeric, we can see the correlation between the diferent variables:

```python
fig = plt.figure(figsize =(12,10))
np.set_printoptions(precision=2)
sns.heatmap(data_train.corr(), annot = True, fmt=".2f")
```

As can be seen in the heatmap, the variables with are correlated are:
- Front camera Mega pixels and Principal camera Mega pixels.
- Pixels height and Pixel width.
- Screen height and Screen weight.
- three_g and four_g.
- Price range and ram.

The price range depends highly of the ram. Let's see graphically:

```python
sns.pairplot(data_train[['price_range','ram']])
```

We can see, that the price range is a categorical variable, so the point biserial correlation coefficient is a special case of Pearsonâ€™s correlation coefficient, which related categorial and numerical variables.
Let's see it:

```python
import scipy.stats as stats

#calculate point-biserial correlation
stats.pointbiserialr(data_train['price_range'], data_train['ram'])
```

Since the p-value of this correlation is less than 0.05, this correlation is statistically significant. 
The more ram the mobile phone has, the higher the price_range.

The fc and pc, sc_h and sc_w, as well as, three_g and four_g seems categorical data as we can see in the pairplot. Therefore, we use spearman coefficient to see the correlation between these variables. 

```python
sns.pairplot(data_train[['fc','pc']])
sns.pairplot(data_train[['three_g','four_g']])
sns.pairplot(data_train[['sc_h','sc_w']])
```

```python
stats.spearmanr(data_train['fc'], data_train['pc'])
```

```python
stats.spearmanr(data_train['three_g'], data_train['four_g'])
```

```python
stats.spearmanr(data_train['sc_w'], data_train['sc_h'])
```

In both relationship, the p-values is lower than 0.05 so these correlations are statistically significant. 

```python
stats.pearsonr(data_train['px_height'], data_train['px_width'])
```
