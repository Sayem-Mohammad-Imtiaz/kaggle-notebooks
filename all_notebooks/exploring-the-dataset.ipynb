```python
import datetime
```

```python
datetime.datetime.strptime("2:06:04", "%H:%M:%S")
```

```python
def hmsToDatetime(x):
    return datetime.datetime.strptime(x, "%H:%M:%S").time()
```

```python
hmsToDatetime("2:06:04")
```

```python
checkout = df_bike["checkout_time"].apply(hmsToDatetime)
```

```python
checkout.head()
```

```python
plt.hist(checkout.values)
```
