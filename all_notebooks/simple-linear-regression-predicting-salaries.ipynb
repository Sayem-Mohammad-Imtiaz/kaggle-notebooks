```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#importing dataset and declaring independent and dependent variable
#the dependent variable is in last column, hence y is assigned with [:,-1] (meaning all rows of last column)
ds = pd.read_csv('../input/salary/Salary.csv')
x = ds.iloc[:,:-1].values
y = ds.iloc[:,-1].values
```

```python
#checking our dataset
ds.head(10)
```

```python
#checking for missing values
ds.isnull().any()
```

no categorical data and no missing values

```python
#splitting dataset in training and testing set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
#training the model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
#visualising training test results
plt.scatter(x_train,y_train,c = 'red')
plt.plot(x_train, lr.predict(x_train))
plt.xlabel('number of years')
plt.ylabel('salary')
plt.title('visualising training test results')
```

```python
#visualising test set results
plt.scatter(x_test,y_test,c = 'red')
plt.plot(x_train, lr.predict(x_train))
plt.xlabel('number of years')
plt.ylabel('salary')
plt.title('visualising test set results')
```

```python
#finding equation of linear regression
c = lr.intercept_
m = lr.coef_
print('equation of linear regression - y = {}x + {}'.format(m,c))
```

```python
#getting accuracy of the model
from sklearn.metrics import r2_score
r2_score(y_test,lr.predict(x_test))
```

**THE ACCURACY FOR THE MODEL CAME OUT TO BE 97% WHICH IS PRETTY GOOD.**
