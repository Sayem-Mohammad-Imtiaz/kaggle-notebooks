**2. Visualization**

Matplot is a python library that help us to plot data. The easiest and basic plots are line, scatter and histogram plots. I want to show you visualization using all plot type. However there is a problem about dataset columns. Among dataset column names don't have to be space so we should edit it.

```python
liste = ["country", "region", "happiness_rank", "happiness_score", "standard_error", "economy", "family", "health", "freedom", "trust", "generosity", "dystopia_residual"]
data.columns = liste
print(data.columns)
```

```python
data.happiness_score.plot(kind="line", color = "red", linewidth = 1, alpha = 1, grid = True, linestyle = ":", figsize = (10,10))
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Happiness Rank Line Plot")
plt.show()
```

```python
data.plot(kind = "scatter", x = "health", y = "economy", alpha = 0.7, color = "red", figsize = (10,10))
plt.xlabel("health")
plt.ylabel("economy")
plt.title("Scatter Plot")
plt.show()
```

```python
data.freedom.plot(kind = "hist", bins = 70, figsize = (10,10))
plt.title("Histogram")
plt.show()
```

**3. Data Filtering**

In this part, we will try to filter data in our dataset. Let's start practice.

We can write to console only West Europe countries in our dataset.

```python
west_europe = data[data.region == "Western Europe"]
print(west_europe.country)
```

We could write to console these countries using 'for loop'. 

```python
for each in west_europe.country:
    print(each)
```

We could filter to 'and operation' between two data .

```python
filtered_data = data[np.logical_and(data.economy > 1, data.region == "Western Europe")]
print(filtered_data.country)
```
