### About the Notebook
Thus, in this notebook I will present efficient and fast method for searching literature, articles, medical reports, and rank a drug based on the sentence sentiment. Meaning, I will check if the drug was related in a positive way or a negative way.
Further work will analyze each one of the paragraph and will automatically offer the circumstances for the chosen sentiment.

### The Method
I used to this end the idea presented in Fine-grained Sentiment Classification using BERT, including their source code published.
The idea is to use BERT to fine tune it to solve fine-grained sentiment classification task, which is more accurate.
BERT is very efficient in fine tuning, since it support adding just one classification layer for the specific task. 
The following figure shows the BERT architrecture. Further reading can be done [here](https://towardsdatascience.com/bert-explained-state-of-the-art-language-model-for-nlp-f8b21a9b6270?gi=e561367aa480).
