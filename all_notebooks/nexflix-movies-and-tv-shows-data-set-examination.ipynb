There are a lot of movies and Tv shows in data set. But, i do not want to work all. So, i want to obtain data set for myself with first and last 10 rows.

```python
# to concatenate first and last 10 rows, give name to write clean code
data1 = data.head(10)
data2 = data.tail(10)
conc_data_row = pd.concat([data1,data2],axis=0,ignore_index=True) #axis=0 means that will add as rows.
conc_data_row
```

```python
data.info() # to get info about data set
```

Let's take 'director' data. Data set includes 4265 directors. But, rest of them (1969) is missing.



```python
data['director'].value_counts(dropna=False) #to see content of director (NaN->1969)
```

Let's fill NaN values as **'empty'** for our data set **conc_data_row**.

```python
conc_data_row['director'].fillna('empty',inplace=True) #when we give true for inplace, we do not assign this to new variable. 
```

When we run code above, we do not see any outputs. But, we should check does it work or not. In order to do that, we should use **assert** function. 

```python
assert conc_data_row['director'].notnull().all() #this return nothing because it is true.
conc_data_row #when we call our data set, we can see missing director values are 'empty' right now. 
```

We filled NaN director values as 'empty'. Now, i will use list comprehension method to see this empty values. It's easy and useful method.

```python
listDirector = conc_data_row['director']
listDirectorEdit = [print('There is NO info about director.') if i=='empty' else print('There is info about its director.') for i in listDirector]
```
