```python
# Plotting State/UTs vs Total Cases
plt.figure(figsize=(20,8))
Plot = sns.barplot(x="State/UTs", y="Total Cases", data=Coviddata)
```

```python
# Top 5 active cases in the country
top_5 = Coviddata.sort_values(by=['Active'], ascending=False).head()

### Generate a Barplot
plt.figure(figsize=(15,5))
Active = sns.barplot(top_5['Active'], top_5['State/UTs'], color = 'red', label='Confirmed')

```

```python
# Top 5 total cases in the country
top_5 = Coviddata.sort_values(by=['Total Cases'], ascending=False).head()

plt.figure(figsize=(15,5))
Total = sns.barplot(top_5['Total Cases'], top_5['State/UTs'], color = 'red', label='Total Cases')
```
