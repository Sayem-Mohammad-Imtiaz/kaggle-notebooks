```python
import numpy as np
import pandas as pd
import nltk
import re
from nltk.corpus import stopwords
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
Encode = preprocessing.LabelEncoder()
from sklearn.manifold import TSNE
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import metrics

```

```python
df = pd.read_csv('../input/deepnlp/Sheet_1.csv', encoding='latin-1')
df.head()
```

```python
df = df.drop(["Unnamed: 3", "Unnamed: 4", "Unnamed: 5", "Unnamed: 6","Unnamed: 7"], axis = 1)
df.head()
```

```python
df['class'].value_counts()
```

```python
def wordcloud(dataframe):
    stopwords = set(STOPWORDS)
    wordcloud = WordCloud(background_color="white",stopwords=stopwords).generate(" ".join([i for i in dataframe.str.upper()]))
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.title("bag_composition")

wordcloud(df['response_text'])
```

```python
x = df.response_text
y = df.Label
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=28)

```

```python
NB = MultinomialNB()
tfid = TfidfVectorizer()
vect = CountVectorizer()
```

```python
y_train.head()
```

```python
x_train = vect.fit_transform(x_train)
x_test = vect.transform(x_test)
```

```python
NB.fit(x_train,y_train)
```

```python
y_predict = NB.predict(x_test)
```

```python
y_predict
```

```python
metrics.accuracy_score(y_test, y_predict)
```

```python
rf = RandomForestClassifier(max_depth=10, max_features=10)
rf.fit(x_train,y_train)
rfpredict = rf.predict(x_test)
metrics.accuracy_score(y_test, rfpredict)
```
