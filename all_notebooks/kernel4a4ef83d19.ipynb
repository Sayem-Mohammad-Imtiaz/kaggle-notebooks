```python
df = pd.read_csv("/kaggle/input/ph-recognition/ph-data.csv") # load train data frame in pandas 
```

```python
df.head()
```

```python
ax = sns.pairplot(df)
```

```python
list1 = df.label
phbalance = []
for i in  list1:
    if i < 7:
        phbalance.append(0)
    elif i > 7: 
        phbalance.append(1)
    else :
        phbalance.append(2)
    
```

```python
df['phbalance'] = phbalance
```

```python
df.head()
```

```python
ax = sns.countplot(data = df, x = phbalance)
```

```python
import plotly.express as px
fig = px.scatter_3d(df, x='red', y='green', z='blue',  
              color='label', symbol ='phbalance' )
fig.show()
```

```python
from sklearn.svm import SVC
clf = SVC(gamma='auto')
from sklearn.model_selection import train_test_split
```

```python
x = df[['blue','green', 'red',]]
x.head()
```

```python
y = df['phbalance']
y.head()
```

```python
#Splitting the data into train and test
X_train, X_test, y_train, y_test = train_test_split(x,y , test_size=0.25, random_state=True)
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
clf.score(X_test,y_test)
```
