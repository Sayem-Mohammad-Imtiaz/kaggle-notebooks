```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('/kaggle/input/housesalesprediction/kc_house_data.csv')
```

**Data Preprocessing**

```python
#Check if exsits na column
df.columns[df.isna().any()]
```

```python
%matplotlib inline
```

```python
corr = df[df.columns].corr()
```

```python
sns.heatmap(corr)
```

```python
 sqft_liv_corr = corr[['sqft_living']]
```

```python
sqft_liv_corr
```

```python
#select features with relative high correlation rate 
seleted_cols = list(sqft_liv_corr[(sqft_liv_corr.sqft_living>0.2)|(sqft_liv_corr.sqft_living<-0.2)].index)
seleted_cols.remove('sqft_living')

```

```python
selected_df = df[seleted_cols]
```

```python
normalized = selected_df.apply(lambda x: (x - np.mean(x)) / (np.max(x) - np.min(x)))
```

```python
sns.heatmap(normalized.corr())
```

**Train & test data**

```python

#y =np.log(1+np.array(df['sqft_living'].values))

y = df['sqft_living'].values
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(normalized.values, y, test_size = 0.2, random_state = 32) 
```

**Train Model**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
reg = LinearRegression().fit(X_train, y_train)
y_train_predict = reg.predict(X_train)
print('Training Performance:','R^2', reg.score(X_train, y_train),'MSE:',mean_squared_error(y_train,y_train_predict))


```

**Test model**

```python
y_predict = reg.predict(X_test)
```

**Evaluate**

```python
print('Testing Performance:', 'R^2:',reg.score(X_test, y_test),
      'MSE:',mean_squared_error(y_test,y_predict))
```
