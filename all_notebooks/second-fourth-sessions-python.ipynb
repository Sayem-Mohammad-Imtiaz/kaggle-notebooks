```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


#1st Question

df=pd.read_csv("../input/hotel-booking/hotel_booking.csv") 
```

```python
#2nd Question

df.info()
```

```python
#3rd Question

number_of_rows = len(df.index)
print(number_of_rows)
```

```python
#4th Question

#This line denotes missing data in every column
df.isna().sum()

#This line indicates that which column has the maximum number of missing data
df.count().idxmin()
```

```python
#5th Question

df.drop('company', axis=1, inplace=True)
df.head()
```

```python
#6th Question

#Calculating the number of booked passengers
Booked_Passengers = (df['adults'] + df['children'] + df['babies']) - df['is_canceled'];

#Inserting the column of 'Booked_Passengers' as a new column into the previous DataFrame
df['Booked_Passengers'] = Booked_Passengers;

#Creating a new DataFrame
NewDataFrame = df[['country','Booked_Passengers']]

#Aggregating the number of the passengers for each country
A = NewDataFrame.groupby('country').sum()

#Sorting the values in a non-ascending (descending) order
B = A.sort_values('Booked_Passengers', ascending=False)

#Selecting five countries with a higher number of passengers
C = B[0:5]
print(C)
```

```python
#7th Question

D = df[['name','adr']].groupby('name').sum().sort_values('adr', ascending=False) [0:1]
print(D)
#[max(df.adr),df.name[max(df.adr)]]
```

```python
#8th Question

E = round(df['adr'].mean(), 2)
print(E)
```

```python
#9th Question

#Calculating the number of accommodation nights
Number_of_accommodation_nights = (df['stays_in_week_nights'] + df['stays_in_weekend_nights']) ;
F = round(Number_of_accommodation_nights.mean(), 2)
print(F)
```

```python
#11th Question

#Spliting the column of name into two columns of first name and last name
H = df.name.str.split(" ",expand=True,)

#Selecting the last name
H[1]

#Finding the 5 most common last names
H[1].value_counts()[0:5]

#df['name'].value_counts()[0:5]
```

```python
#12th Question

#Aggregating the number of babies and children
I = df['babies'] + df['children']

#Inserting the column of 'I' as a new column into the previous DataFrame
df['Total_number_of_babies_and_children'] = I;

#Creating a new DataFrame
NovelDataFrame = df[['name','Total_number_of_babies_and_children']]

#Identifying individuals with a higher number of babies and children
J = NovelDataFrame[['name','Total_number_of_babies_and_children']].sort_values('Total_number_of_babies_and_children', ascending=False)[0:2]

print(J)
```

```python
#13th Question

#Spliting the area code as a new column
Code = df['phone-number'].apply(lambda area_code:area_code[:3])

Code.value_counts()[0:3]
```

**The below code provides information about relation between "Number of accommodation nights" and "Booked Passengers" for each customer type using scatter plot.**

```python
Number_of_accommodation_nights = (df['stays_in_week_nights'] + df['stays_in_weekend_nights']) ;
df['Number_of_accommodation_nights'] = Number_of_accommodation_nights;

plt.figure(figsize=(4, 4), dpi=200)
sns.scatterplot(data=df, x='Number_of_accommodation_nights', y='Booked_Passengers', hue='customer_type',style='customer_type', palette='flare')
```

**Insight:** Overall, the number of booked passengers who have stayed more than 20 nights was less than 10 persons. In other words, it is more probable that passengers who come in a less populated group may stay more nights.
