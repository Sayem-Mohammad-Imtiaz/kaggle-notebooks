```python
import plotly.offline as pyo

import matplotlib.pyplot as plt
import seaborn as sns

from wordcloud import WordCloud, STOPWORDS
from PIL import Image

import plotly.graph_objs as go
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

pd.options.display.max_columns = None

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

# First look at data

```python
df = pd.read_csv('/kaggle/input/cyber-incidents-up-to-2020/cyber-operations-incidents.csv')
df.head()
```

# The type of attacks

Most of the recorded type of cyber attack is espionage.

```python
order = df['Type'].value_counts().index

fig, ax = plt.subplots(1, 1, figsize=(15, 9))
sns.countplot(data=df, x='Type', order=order, palette='winter')
ax.grid(axis='y', linewidth=0.5, alpha=0.25, color='red')
ax.set(xlabel='Type of cyber attack')
ax.set(ylabel='Count of attacks')
ax.set_facecolor('black')

plt.title('Count of various types of cyber attacks', fontsize = 20, c='black')
plt.show()
```

```python
background = np.array(Image.open('../input/white-background/black.png'))
```

# Wordcloud: All data

```python
word_string=" ".join(df['Description'].str.lower())
wordcloud = WordCloud(stopwords=STOPWORDS, mask=background, colormap='winter').generate(word_string)

fig, ax = plt.subplots(1, 1, figsize=(15, 9))
plt.clf()
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = pd.DatetimeIndex(df['Date']).year
```

# Wordcloud - year wise

*"Cyber threat actors are states, groups, or individuals who, with malicious intent, aim to take advantage of vulnerabilities, low cyber security awareness, or technological developments to gain unauthorized access to information systems in order to access or otherwise affect victims' data, devices, systems, and networks"*

```python
fig, ax = plt.subplots(3,2,figsize=(15, 35))
[axi.set_axis_off() for axi in ax.ravel()]
i = 1
    
for year in df['Year'].unique()[:-1]:
    
    word_string=" ".join(df[df['Year']==year]['Description'].str.lower())
    ax = fig.add_subplot(8,2,i)
    ax.imshow(WordCloud(stopwords=STOPWORDS, mask=background, colormap='winter').generate(word_string))
    ax.title.set_text(year)
    plt.axis('off')
    i = i + 1

plt.show()
```
