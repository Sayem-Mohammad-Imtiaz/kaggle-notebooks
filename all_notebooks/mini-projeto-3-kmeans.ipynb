# Mini-projeto 3: Clusterização com Kmeans

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

colleges = pd.read_csv('/kaggle/input/us-news-and-world-reports-college-data/College.csv',index_col=0)
colleges.describe()
```

Criaremos uma nova coluna, a taxa de gastos. Ela mede quanto é gasto por aluno sobre quanto a instituição recebe de taxa(Out-of-state).

```python
colleges['Expend Ratio'] = colleges['Expend']/(colleges['Outstate'])
colleges.describe()['Expend Ratio']
```

Podemos notar que a instituição média tem um Expende Ratio menor que um, então ela gasta menos do que ela recebe por Out-of-state tuition.

```python
sns.set_style('darkgrid')

def plot_graph(column):
    grid = sns.FacetGrid(colleges,hue='Private', height=6, aspect=1.5)
    grid = grid.map(plt.hist, column, bins=20, alpha=0.7)
    plt.legend(labels=['Privadas','Públicas'])

plot_graph('Outstate')
```

Gráfico 1: Quantidade de instituições por valor de Out-of-state tuition. Divididas em pública em privada.
Podemos notar que intituições privadas posuem um Out-of-state tuition maior que as públicas.

```python
plot_graph('Grad.Rate')
```

Gráfico 2: Quantidade de instituições por taxa de graduação. Divididas em pública em privada.
Podemos notar que intituições privadas posuem uma taxa de graduação maior que as públicas.

Como notado acima, uma instituição tem taxa de graduação de mais de 100%. Desas forma, vamos normalizar esse dado:

```python
def fixGradRate(rate):
    if rate > 100:
        return 100
    return rate
colleges['Grad.Rate'] = colleges['Grad.Rate'].apply(fixGradRate)
plot_graph('Grad.Rate')
```

```python
plot_graph('Expend Ratio')
```

Utilizando Kmeans:

Iremos dividir os clusters em: Instituições Privadas e Públicas. Para isso, vamos traduzir a string 'Yes', na coluna 'Private' para o cluster 1 e o resto vai para o cluster 1.

```python
def translateToCluster(cluster):
    if cluster=='Yes':
        return 1
    else:
        return 0
colleges['Private'] = colleges['Private'].apply(translateToCluster)
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=2, random_state=10)


y = colleges['Private']
x = colleges.drop(['Private', 'Expend Ratio'],axis=1)

from sklearn.model_selection import train_test_split
# Subset de treino e teste:
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.5, random_state=15)
# Treinando:
km.fit(x_train)
```

```python
predicted = km.predict(x_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,predicted))
```
