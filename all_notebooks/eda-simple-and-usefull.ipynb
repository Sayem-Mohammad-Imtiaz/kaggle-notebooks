# Most Popular Platform, Publisher And Year:

```python
most_platform = sort_global_sale['Platform'].value_counts()[:16]
```

```python
def change_platform(x):
    if x in most_platform.index:
        return x
    else:
        return np.nan
```

```python
video_game['Platform'] = video_game['Platform'].apply(change_platform)
```

```python
video_game.dropna(inplace=True)
```

```python
video_game
```

```python
most_publisher = video_game['Publisher'].value_counts()[:10]
```

```python
def change_publisher(x):
    if x in most_publisher.index:
        return x
    else:
        return np.nan
```

```python
video_game['Publisher'] = video_game['Publisher'].apply(change_publisher)
```

```python
video_game.dropna(inplace=True)
```

```python
video_game['Year'] = video_game['Year'].astype('int32')
```

```python
most_year = video_game['Year'].value_counts()[:15]
most_year.index
```

```python
def change_year(x):
    if x in most_year.index:
        return x
    else:
        return np.nan
```

```python
video_game['Year'] = video_game['Year'].apply(change_year)
```

```python
video_game.dropna(inplace=True)
```

```python
a = video_game.pivot_table(index='Publisher', values='Global_Sales', columns='Year', aggfunc='max', fill_value='0')
```

```python
a.plot(kind='barh', figsize = (25, 25))
```
