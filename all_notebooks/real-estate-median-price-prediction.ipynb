## Real Estate Median Price Prediction  

Given *data about real estate in Boston*, let's try to predict the **median values** of a given home.  
  
We will use a random forest regressor to make our predictions.

# Getting Started

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestRegressor
```

```python
data = pd.read_csv('../input/real-estate-dataset/data.csv')
```

```python
data
```

```python
data.info()
```

# Preprocessing

```python
data['RM'] = data['RM'].fillna(data['RM'].mean())
```

```python
print("Total missing values:", data.isna().sum().sum())
```

```python
data
```

```python
y = data['MEDV'].copy()
X = data.drop('MEDV', axis=1).copy()
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

# Training

```python
no_bootstrap_model = RandomForestRegressor(n_estimators=100, criterion='mse', bootstrap=False)
no_bootstrap_model.fit(X_train, y_train)
```

```python
bootstrap_model = RandomForestRegressor(n_estimators=100, criterion='mse', bootstrap=True)
bootstrap_model.fit(X_train, y_train)
```

# Results

```python
print("R^2 without bootstrapping:", no_bootstrap_model.score(X_test, y_test))
print("   R^2 with bootstrapping:", bootstrap_model.score(X_test, y_test))
```
