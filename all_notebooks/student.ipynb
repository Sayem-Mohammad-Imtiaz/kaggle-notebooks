```python
stu_por=pd.read_csv("../input/acamedicperfomance/student_language.csv",sep=";")
stu_mat=pd.read_csv("../input/acamedicperfomance/student_math.csv",sep=";")
```

```python
stu_por.insert(loc=1, column='Course', value='Portuguese')
stu_mat.insert(loc=1, column='Course', value='Math')
```

```python
stu_mat['sex']=stu_mat['sex'].replace('F','Female')
stu_mat['sex']=stu_mat['sex'].replace('M','Male')
```

```python
stu_por=stu_por.replace('other',np.nan)
stu_mat=stu_mat.replace('other',np.nan)
stu_por_GP=stu_por[stu_por['school']=='GP']
stu_por_MS=stu_por[stu_por['school']=='MS']
stu_mat_GP=stu_mat[stu_mat['school']=='GP']
stu_mat_MS=stu_mat[stu_mat['school']=='MS']
```

```python
stu_por_GP.to_csv('stu_por_GP.csv',index=False)
stu_por_MS.to_csv('stu_por_MS.csv',index=False)
stu_mat_GP.to_csv('stu_mat_GP.csv',index=False)
stu_mat_MS.to_csv('stu_mat_MS.csv',index=False)
```
