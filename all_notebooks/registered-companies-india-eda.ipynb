```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../input/monthly-list-of-legal-entities-registered-in-india/2021_april_registered_companies.csv")
```

```python
sns.heatmap(df.corr())
```

```python
fig,ax=plt.subplots(1,2,figsize=(20,5))
df['class'].value_counts().plot.pie(explode=None,ax=ax[0],autopct='%1.1f%%',shadow=True)
ax[0].set_title("Count of Class of Companies")
ax[0].set_ylabel("Count")

sns.countplot("category",data=df,order=df['category'].value_counts().index)
ax[1].set_title("Count of types of Categories Companies are Listed")
plt.show()
```

```python
sns.set()

ax1 = df['state'].value_counts().plot(kind='barh', figsize=(20,10), fontsize=13, color = ['red', 'orange', 'yellow','green','blue',
                                                                                        'indigo','violet','purple','pink','silver',
                                                                                        'gold','beige','brown','grey','black','white'])

for tick in ax1.get_xticklabels():
    tick.set_rotation(0)

ax1.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=15)

plt.show()
```

```python
sns.set()

ax1 = df['company_type'].value_counts().plot(kind='bar', figsize=(20,10), fontsize=13, color = ['red', 'orange', 'yellow','green','blue'])

for tick in ax1.get_xticklabels():
    tick.set_rotation(0)

ax1.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=15)

plt.show()
```
