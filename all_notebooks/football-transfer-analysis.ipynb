# 0. Add the features

```python
df['Age']=2021-df['Born']
df['fromto_country']=df['From(Country)'].str.cat(df['To(Country)'], sep=' to ')
df['fromto_team']=df['From(Club)'].str.cat(df['To(Club)'], sep=' to ')
```

# 6. Total Fee by Origin

# 7. How much did team get ?

```python
df.groupby('From(Club)')['Fee(€ mln)'].sum().sort_values(ascending=False).plot.bar(figsize=(8,8))
```

# 8. How much did team pay ?

```python
df.groupby('To(Club)')['Fee(€ mln)'].sum().sort_values(ascending=False).plot.bar(figsize=(8,8))
```

# 9. Net Revenue (Plus - Minus) by Club

```python
df1=pd.DataFrame(df.groupby('From(Club)')['Fee(€ mln)'].sum())
df2=pd.DataFrame(df.groupby('To(Club)')['Fee(€ mln)'].sum())
df1=df1.rename(columns={'Fee(€ mln)': 'Plus'})
df2=df2.rename(columns={'Fee(€ mln)': 'Minus'})
df1=df1.reset_index()
df2=df2.reset_index()
df1=df1.rename(columns={'From(Club)': 'Club'})
df2=df2.rename(columns={'To(Club)': 'Club'})
df_pm=pd.merge(df1, df2, on='Club', how='outer')
df_pm=df_pm.fillna(0)
df_pm=df_pm.set_index('Club')
df_pm['Net']=df_pm['Plus']-df_pm['Minus']
```

```python
df_pm['Net'].sort_values(ascending=False).plot.bar(figsize=(10,10))
```
