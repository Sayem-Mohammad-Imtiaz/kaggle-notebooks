```python
calls_df['timeStamp'] =pd.to_datetime(calls_df['timeStamp'])
```

```python
time = calls_df['timeStamp'].iloc[0]
```

```python
time.year
```

```python
calls_df['Hour'] = calls_df['timeStamp'].apply(lambda time: time.hour)
calls_df['Day of week'] = calls_df['timeStamp'].apply(lambda time: time.dayofweek)
calls_df['Month'] = calls_df['timeStamp'].apply(lambda time: time.month)
```

```python
calls_df.head()
```

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'}
calls_df['Day of week'] = calls_df['Day of week'].map(dmap)
```

```python
calls_df.head()
```

```python
##### Day of week against Reason
```

```python
sns.countplot(x='Day of week', data=calls_df,hue='Reason')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
```

Friday has the most number of calls

```python
sns.countplot(x='Month', data=calls_df,hue='Reason')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0)
```

 January has the most calls
