```python
df = pd.read_csv('/kaggle/input/oranges-vs-grapefruit/citrus.csv')
df.head()
```

```python
df.info()
df.describe()
```

```python
plt.figure(figsize=(10,6))
df[df['name']=='orange']['diameter'].hist(alpha=0.5,color='blue',bins=50)
df[df['name']=='grapefruit']['diameter'].hist(alpha=0.5,color='red',bins=50)
plt.legend()
plt.xlabel('diameter')
plt.ylabel('name')
```

```python
sns.heatmap(df.corr(),cmap='coolwarm')
plt.title('citrus.corr()')
```

```python
final_data = pd.get_dummies(df,columns=['name'],drop_first=True)
final_data.head()
```

```python
X=final_data.drop(['name_orange'],axis=1)
y=final_data['name_orange']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)

dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
predictions = dtree.predict(X_test)

print(classification_report(y_test,predictions))
print('conf matrix:\n',confusion_matrix(y_test,predictions))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(max_leaf_nodes = 100, splitter = 'random')
dtree.fit(X_train,y_train)
predictions = dtree.predict(X_test)

print(classification_report(y_test,predictions))
print('conf matrix:\n',confusion_matrix(y_test,predictions))
```

```python

rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
predictions = rfc.predict(X_test)
print(classification_report(y_test,predictions))

print('conf matrix:\n',confusion_matrix(y_test,predictions))
```

```python
rfc = RandomForestClassifier(n_estimators=1000,max_features = 'sqrt',max_depth=10, min_samples_split=10, min_samples_leaf=2, bootstrap=False)
rfc.fit(X_train,y_train)
predictions = rfc.predict(X_test)
print(classification_report(y_test,predictions))

print('conf matrix:\n',confusion_matrix(y_test,predictions))
```
