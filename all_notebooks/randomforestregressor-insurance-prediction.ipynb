```python
X, y = df.drop('charges', axis=1), df.charges
print(X.shape, y.shape)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
columnTransformer = ColumnTransformer([('encoder', OneHotEncoder(), [1,4,5])], remainder='passthrough') 
X = columnTransformer.fit_transform(X)
X[0:5]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
print(y_train.shape, y_test.shape)
```

```python
st = StandardScaler()
X_train = st.fit_transform(X_train)
X_test = st.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = { 
    'n_estimators': [200, 500],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :["mse", "mae"]
}

rfr=RandomForestRegressor(random_state=42, verbose=1, n_jobs=-1)

CV_rfr = GridSearchCV(estimator=rfr, param_grid=param_grid, cv= 5, verbose=True)
CV_rfr.fit(X_train, y_train)
```

```python
CV_rfr.best_params_
```

```python
model = RandomForestRegressor(criterion= 'mse',
 max_depth= 4,
 max_features= 'auto',
 n_estimators= 500)

model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
pred
```

```python
mean_squared_error(y_test, pred)
```

```python
r2_score(y_test, pred)
```
