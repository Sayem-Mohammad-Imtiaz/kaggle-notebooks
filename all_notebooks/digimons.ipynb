# **Distribution within stages.**

```python
stage_count = dmlist.stage.value_counts()
fig = px.bar(stage_count, x=stage_count.index, y=stage_count, labels={'y':'count', 'index':'stage'}, )
fig.show()
```

# Different types of Digimons.

```python
dmpie = dmlist.groupby('type')['digimon'].count()
dmpie
```

```python
type_col = ['type', 'amount']
type_data = [
    ['Data', 60],
    ['Free', 37],
    ['Vaccine', 70],
    ['Virus', 82]
]
type_tab = pd.DataFrame(data=type_data, columns=type_col)
fig = px.pie(type_tab, values='amount', names='type', color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```
