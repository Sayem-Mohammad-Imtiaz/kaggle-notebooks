```python
df_questions = pd.read_csv('../input/ideal-student-life-survey/survey_questions_meta.csv')
df_questions.shape
```

# **Dividing Questions into 5 lists**

```python
cols = list(df_questions.columns)
q1 = []
q2 = []
q3 = []
q4 = []
q5 = []

for col in cols:
    if "Q1" in col:
        q1.append(col)
    elif "Q2" in col:
        q2.append(col)
    elif "Q3" in col:
        q3.append(col)
    elif "Q4" in col:
        q4.append(col)
    elif "Q5" in col:
        q5.append(col)
    
```

```python
def plot_pie(responses):
    cols = list(responses)
    for col in cols:
        plt.style.use('seaborn')
        question_data = df_questions[col].value_counts()
        question_data.plot(kind='pie', autopct='%.2f')
        plt.title(f'{col}')
        plt.ylabel('')
        plt.xlabel(f'Total Responses: {question_data.sum()}')
        plt.show()
        print('\n\n')
```

# **Reponses to Q1 Questions (Volunteered for?)**

```python
plot_pie(q1)
```

# **Reponses to Q2 Questions (Participated in?)**

```python
plot_pie(q2)
```

# **Reponses to Q3 Questions (Interested in?)**

```python
plot_pie(q3)
```

# **Reponses to Q4 Questions (Passionate about?)**

```python
plot_pie(q4)
```

# **Reponses to Q5 Questions (Stressed about?)**

```python
plot_pie(q5)
```
