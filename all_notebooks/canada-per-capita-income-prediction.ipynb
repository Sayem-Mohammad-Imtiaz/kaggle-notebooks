```python
df.head()

df=df.rename({'per capita income (US$)': 'Per_capital_income'}, axis=1)  

df.head()

plt.scatter(df.year,df.Per_capital_income)

reg=linear_model.LinearRegression()

reg.fit(df[['year']],df.Per_capital_income)

reg.predict([[2020]])

plt.scatter(df.year,df.Per_capital_income,linestyle=":",color='red',marker='+')
plt.plot(df.year,reg.predict(df[['year']]),linestyle=":",color='blue')

reg.coef_

reg.intercept_
```
