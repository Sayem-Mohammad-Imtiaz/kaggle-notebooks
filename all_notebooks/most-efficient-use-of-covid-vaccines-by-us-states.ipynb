# Covid-19 Vaccination efficient use in last week

This notebook contains code to identify US states which are using their vaccine supplies most efficiently for past 7 days:

* A list of the top 10 and bottom 10 states which are using their vaccine supplies most efficiently

New Mexico is using it in the best way and Marshall Islands is having minimum share_doses_used

```python
#Importing libraries
import pandas as pd
import numpy as np
import datetime
```

```python
#Loading the dataset
df_us_covid19=pd.read_csv('../input/usa-covid19-vaccinations/us_state_vaccinations.csv')

#Checking top 5 rows
df_us_covid19.head()
```

```python
df_us_covid19.info()
```

```python
#Unique values of location
df_us_covid19['location'].unique()
```

It seems below are not US states but are also part of location column, dropping it for the analysis. 'Bureau of Prisons', 'Dept of Defense', 'United States', 'Veterans Health'

```python
#Dropping Non States entries
non_states = df_us_covid19['location'].isin(['Bureau of Prisons', 'Dept of Defense', 'United States', 'Veterans Health'])
df_us_covid19 = df_us_covid19[~non_states]
```

```python
#Except date and location all other fields are Numbers, hence replacing Null with zero
df_us_covid19=df_us_covid19.fillna(0)
```

```python
#Max date in the dataset and calculating last 7 days in data
df_us_covid19['date_modified']=pd.to_datetime(df_us_covid19['date'])
max_date=df_us_covid19['date_modified'].max()
print(max_date)
min_date=max_date-pd.to_timedelta("6day")
print(min_date)
```

```python
#Filtering dataset for last 7 days
df_us_covid19_7days=df_us_covid19.loc[(df_us_covid19['date_modified'] >= min_date) & (df_us_covid19['date_modified'] <= max_date)]
```

```python
#Checking Min and Max of filtered dataset
print(df_us_covid19_7days['date_modified'].max())
print(df_us_covid19_7days['date_modified'].min())
```

```python
df_A=df_us_covid19_7days.groupby('location')['share_doses_used'].sum().reset_index()
df_A.sort_values('share_doses_used',ascending=False).head(10)
```

```python
df_B=df_us_covid19_7days.groupby('location')['share_doses_used'].sum().reset_index()
df_B.sort_values('share_doses_used',ascending=False).tail(10)
```

**New Mexico is using it in the best way and Marshall Islands is having minimum share_doses_used**
