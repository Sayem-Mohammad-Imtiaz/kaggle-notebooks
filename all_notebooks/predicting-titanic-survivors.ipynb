```python
#get the trainset
train=pd.read_csv('/kaggle/input/titanic/train.csv')
# show the head
train.head()
```

Get the percentage of survivors by Gender

```python
#for men 
men=train['Survived'][train.Sex=='male']
percentage =sum(men)/len(men)*100
print('the rate of survived men is {} %'.format(round(percentage,2)))
```

```python
#for women 
women=train['Survived'][train.Sex=='female']
percentage = sum(women)/len(women)*100
print('the rate of survived women is {} %'.format(round(percentage,2)))
```

As we can see the gender matters here !

```python
# Visualize percenteges by sex
sb.set_context('poster')
sb.set_style('ticks')
sb.barplot(x='Sex',y='Survived',data=train)     
sb.despine()     
```

```python
#  Count by Gender 

sb.set_context('poster')
sb.set_style('ticks')
sb.countplot(x='Survived',hue='Sex',data=train)  
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
# Visulise how many survived by Class (High, Middle ,Low )
sb.barplot(x='Pclass',y='Survived',hue='Sex',data=train)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
#Visulize by Age 
sb.barplot(x='Survived',y='Age',hue='Sex',data=train)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

Train Test Split

```python
from sklearn.model_selection import train_test_split
#get target column
y = train["Survived"]
#get the features .. only  features that matter
features = ["Pclass", "Sex", "SibSp", "Parch"]
X = pd.get_dummies(train[features])


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)

```

```python
# train the model 
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=600, max_depth=5, random_state=1)
rfc.fit(X_train,y_train)
```

```python
# make predicttions on the model
predictions=rfc.predict(X_test)
```

```python
#Let's run some metrics 
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print('Confusion Matrix :\n',confusion_matrix(y_test,predictions))
print('\n')
print('Classification report :\n',classification_report(y_test,predictions))
print('\n')
print('Accuracy score is : {}'.format(round(accuracy_score(y_test,predictions),2)))
```

##### Testing the model on the testset 

```python
#grab the test set
test=pd.read_csv('/kaggle/input/titanic/test.csv')
#show the head of the df 
test.head()
```

#### Make the submission on the testset

```python
features = ["Pclass", "Sex", "SibSp", "Parch"]
X_test = pd.get_dummies(test[features])
test_predictions=rfc.predict(X_test)
#save my submission 
submission= pd.read_csv('../input/titanic/gender_submission.csv')
submission['Survived']=test_predictions
#save the submission
submission.to_csv('submission.csv', index=False)
print(" Submission  successfully saved!")
```
