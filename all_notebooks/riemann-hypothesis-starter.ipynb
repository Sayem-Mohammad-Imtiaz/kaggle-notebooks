# We jsut have to build model that predict next prime number

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.model_selection import train_test_split
```

```python
df_s = df.head(500)
df_s = df_s.replace([np.inf, -np.inf], np.nan)
df_s = df_s.dropna()
X = df_s.drop(columns='prime')
y = df_s.prime
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train,y_train)
y_pred = svm.predict(X_test)
print('Root Mean Square Error is ',np.sqrt(mse(y_test,y_pred)))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X_train,y_train)
y_pred = lr.predict(X_test)
print('RMSE ',np.sqrt(mse(y_test,y_pred)))
```

# As we increasing size of data error also increasing

```python
df_s = df.head(3000)
df_s = df_s.replace([np.inf, -np.inf], np.nan)
df_s = df_s.dropna()
X = df_s.drop(columns='prime')
y = df_s.prime
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(X_train,y_train)
y_pred = svm.predict(X_test)
print('Root Mean Square Error is ',np.sqrt(mse(y_test,y_pred)))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X_train,y_train)
y_pred = lr.predict(X_test)
print('RMSE ',np.sqrt(mse(y_test,y_pred)))
```

```python
import lightgbm as gbm

lgb = gbm.LGBMRegressor()
lgb.fit(X_train,y_train)
y_pred = lgb.predict(X_test)
print('RMSE ',np.sqrt(mse(y_test,y_pred)))
```
