```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

df = pd.read_csv('../input/data.csv')
df.drop(df[df['MSRP'] < 10000].index,inplace = True)
```

```python
print(df['Vehicle Size'].unique())
df['Vehicle Size'] = df['Vehicle Size'].replace(df['Vehicle Size'].unique(), [1, 2, 3])
```

```python
print(df['Vehicle Style'].unique())
df['Vehicle Style'] = df['Vehicle Style'].replace(df['Vehicle Style'].unique(), [1, 3, 2, 4, 8, 8, 5, 6, 6, 7, 7, 7, 5, 6, 5, 6])
```

```python
print(df['Engine Fuel Type'].unique())
df['Engine Fuel Type'] = df['Engine Fuel Type'].replace(df['Engine Fuel Type'].unique(), [2, 2, 2, 2, 1, 3, 2, 4, 2, 2, 2])
```

```python
print(df['Transmission Type'].unique())
df['Transmission Type'] = df['Transmission Type'].replace(df['Transmission Type'].unique(), [1, 2, 2, 2])
```

```python
print(df['Driven_Wheels'].unique())
df['Driven_Wheels'] = df['Driven_Wheels'].replace(df['Driven_Wheels'].unique(), [2, 1, 3, 2])
```

```python
print(df['Market Category'].unique())
df['Market Category']=df['Market Category'].replace(np.nan, 'regular')
df['Market Category'].unique()
for category in df['Market Category'].unique():   
    if 'Luxury' in category:
        df['Market Category']=df['Market Category'].replace(category,1)
    else:
        df['Market Category']=df['Market Category'].replace(category,2)
```

```python
df.groupby('Engine HP').count()['MSRP'].plot()
```

```python
def cut(x):
    if 50 < x <= 200:
        return 1
    elif 200 < x <= 250:
        return 2
    elif 250 < x <= 300:
        return 3
    elif 300 < x <= 350:
        return 4
    elif 350 < x <= 400:
        return 5
    else:
        return 6
```

```python
df['Engine HP Category'] = df['Engine HP'].apply(cut)
df.drop('Engine HP', axis=1, inplace=True)
df.groupby('Engine HP Category').count()['MSRP']
```

```python
df['Name'] = df[['Make','Model']].apply(lambda x: ' '.join(x), axis = 1)
```

```python
df = df.drop_duplicates(subset = 'Name',keep = 'first')
df = df.reset_index(drop = True)
```

```python
df.drop(['Year','Name'],axis=1,inplace=True)
```
