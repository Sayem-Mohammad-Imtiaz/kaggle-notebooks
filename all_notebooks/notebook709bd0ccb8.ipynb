```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas_profiling import ProfileReport
```

```python
data=pd.read_csv("../input/covid19-world-testing-progress/covid-testing.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data['Date']=pd.to_datetime(data['Date'])
```

```python
plt.figure(figsize=(30,7))
sns.countplot(data['Source label'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.style.use('dark_background')
sns.lineplot(x='Date',y='Cumulative total',data=data)
```

```python
data.groupby("ISO code")['Cumulative total'].sum().sort_values(ascending=False).head(20).plot.bar()
```

```python
data.groupby("ISO code")['Cumulative total'].sum().sort_values(ascending=True).head(20).plot.bar()
```

```python
data.groupby("ISO code")['Daily change in cumulative total'].sum().sort_values(ascending=False).head(20).plot.bar()
```

```python
high5 = data[(data["ISO code"]=="USA") |(data["ISO code"]=="IND")|(data["ISO code"]=='RUS') |(data["ISO code"]=='GBR')|(data["ISO code"]=='ITA')]
```

```python
plt.figure(figsize=(20,7))
plt.style.use("ggplot")
sns.lineplot(data=high5,x="Date",y="Cumulative total",marker='o',hue="ISO code")
plt.show()
```

```python
plt.figure(figsize=(12,7))
plt.style.use('fivethirtyeight')
sns.lineplot(x='Date',y='Cumulative total per thousand',data=data)
```

```python
plt.figure(figsize=(12,7))
plt.style.use('fivethirtyeight')
sns.lineplot(x='Date',y='Daily change in cumulative total per thousand',data=data)
```

```python
plt.figure(figsize=(12,7))
plt.style.use('fivethirtyeight')
sns.lineplot(x='Date',y='7-day smoothed daily change',data=data)
```

```python
plt.figure(figsize=(12,7))
plt.style.use('fivethirtyeight')
sns.lineplot(x='Date',y='7-day smoothed daily change per thousand',data=data)
```
