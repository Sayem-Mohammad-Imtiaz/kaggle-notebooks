```python
from sklearn.model_selection import train_test_split
Features_train, Features_test, Prediction_train, Prediction_test = train_test_split(Features, Prediction, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(Features_train,Prediction_train)
```

```python
Actual_Predictions = lm.predict(Features_test)
```

```python
plt.scatter(Prediction_test,Actual_Predictions)
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
print('MAE:', metrics.mean_absolute_error(Prediction_test, Actual_Predictions))
print('MSE:', metrics.mean_squared_error(Prediction_test, Actual_Predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Prediction_test, Actual_Predictions)))

score=r2_score(Prediction_test, Actual_Predictions)
print('score=',+score)
```
