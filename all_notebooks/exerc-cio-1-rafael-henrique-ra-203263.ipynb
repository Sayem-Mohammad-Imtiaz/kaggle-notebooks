A variável qualitativa que mais fez sentido utilizar para criar tabela frequência e frequencia relativa foi a da unidade federal (UF), pois, a variavel cod_municipio_tse é um ID que não se repete, e a variavel nome_municipio se repete (em poucos casos: 4,88%, 272 de 5568) mas são de estados diferentes então não traria informações úteis.

```python
qtdMunicQueRepetem = len(data['nome_municipio'])-len(data['nome_municipio'].unique())
porcMunicQueRepetem =  (qtdMunicQueRepetem/len(data['nome_municipio']))*100
porcMunicQueRepetem
```

```python
table_freq = data['uf'].value_counts().reset_index()
table_freq.columns = ['UF','Freq_Abs']
table_freq['Freq_Rel'] = ((data['uf'].value_counts()/data['uf'].value_counts().sum())*100).reset_index()['uf']

table_freq
```

### Questão 1 - Item C - Representação Gráfica

### Quantidade de Municípios Eleitorais por Estado

```python
fig = plt.figure(figsize=(14, 9))
ax = fig.add_axes([0,0,1,1])


values_x = table_freq['UF']
values_y = table_freq['Freq_Abs']
ax.bar(values_x,values_y)

ax.set_title('Quantidade de Muncípios Eleitorais por Estado')
ax.set_ylabel('Quantidade de Municípios',size=15)
ax.set_xlabel('UF',size=15)

for i, v in enumerate(values_y):
    plt.text(i-0.35, values_y[i] + (max(values_y)/2)*0.05, str(f"{values_y[i]}"), color='black', fontsize=13)

plt.plot



plt.show()
```

### Quantidade de Eleitores por Estado

```python
fig = plt.figure(figsize=(14, 6))
ax = fig.add_axes([0,0,1,1])

dadosAgrupadosPorEstado = data.groupby('uf').sum().sort_values('total_eleitores',ascending=False)

values_x = dadosAgrupadosPorEstado.index.values
values_y = dadosAgrupadosPorEstado['total_eleitores'].values
ax.bar(values_x,values_y)

ax.set_title('Quantidade de Eleitores por Estado',size=17)
ax.set_ylabel('Quantidade de Eleitores',size=15)
ax.set_xlabel('UF',size=15)

values_yticks = np.arange(10000000,50000000,step=5000000)

ax.set_yticks(values_yticks)

for i, v in enumerate(values_y):
    plt.text(i-0.35, values_y[i] + (max(values_y)/2)*0.3, str(f"{values_y[i]}"), color='black', fontsize=13,rotation=45)





plt.show()

print("Observação Importante: A quantidade aprensentada no eixo Y (por Exemplo 3.5) está em notação científica, e ao topo no canto esquerdo diz por qual numero deve ser multiplicado o valor do eixo Y para encontrar o valor real. \n\nNo topo de cada barra encontra-se o valor exato da quantidade de eleitores")
```

### Quanto os eleitores de um estado representam em porcentagem dos eleitores totais?

```python
fig = plt.figure(figsize=(16, 6))
ax = fig.add_axes([0,0,1,1])

dadosAgrupadosPorEstado = data.groupby('uf').sum().sort_values('total_eleitores',ascending=False)

values_x = dadosAgrupadosPorEstado.index.values
values_y = np.around((dadosAgrupadosPorEstado['total_eleitores'].values/data['total_eleitores'].sum())*100,2)
ax.bar(values_x,values_y)

ax.set_title('Porcentagem de Eleitores de cada Estado em relação ao Total de Eleitores',size=20)
ax.set_ylabel('Porcentagem de Eleitores',size=15)
ax.set_xlabel('UF',size=15)

#values_yticks = np.arange(10000000,50000000,step=5000000)

#ax.set_yticks(values_yticks)

for i, v in enumerate(values_y):
    plt.text(i-0.35, values_y[i] + (max(values_y)/2)*0.03, str(f"{values_y[i]}%"), color='black', fontsize=11)

plt.show()
```

### Qual é a faixa de idade que possuem mais eleitores?

```python
fig = plt.figure(figsize=(16, 6))
ax = fig.add_axes([0,0,1,1])

Qtd_idades_Eleitores = data[['f_16','f_17','f_18_20','f_21_24','f_25_34','f_35_44','f_45_59','f_60_69','f_70_79','f_sup_79']].sum().sort_values(ascending=False)

values_x = Qtd_idades_Eleitores.index
values_y_porc = np.around((Qtd_idades_Eleitores.values/data['total_eleitores'].sum())*100,2)
values_y = Qtd_idades_Eleitores
ax.bar(values_x,values_y)

values_yticks = np.arange(10000000,50000000,step=5000000)

ax.set_title('Quantidade de eleitores por faixa de idade',size=20)
ax.set_ylabel('Quantidade de Eleitores',size=15)
ax.set_xlabel('Faixa de Idades dos Eleitores',size=15, labelpad=10)


ax.set_yticks(values_yticks)


for i, v in enumerate(values_y):
    plt.text(i-0.35, values_y[i] + (max(values_y)/2)*0.15, str(f"{values_y[i]}"), color='black', fontsize=13)
    plt.text(i-0.35, values_y[i] + (max(values_y)/2)*0.05, str(f"({values_y_porc[i]}%)"), color='black', fontsize=13)

plt.show()


print("Observação Importante: A quantidade aprensentada no eixo Y (por Exemplo 3.5) está em notação científica, e ao topo no canto esquerdo diz por qual numero deve ser multiplicado o valor do eixo Y para encontrar o valor real. \n\nNo topo de cada barra encontra-se o valor exato da quantidade de eleitores")
```

### Quantidade de Eleitores divididos por gênero e por estado

```python
def autolabel(rects):
    for rect in rects:
        h = rect.get_height()
        ax.text((rect.get_x()+rect.get_width()/2.)+0.2, 1.05*h, '%d'%int(h),
                ha='center', va='bottom',size=11,rotation=45)
```

```python
fig = plt.figure(figsize=(25, 6))
ax = fig.add_axes([0,0,1,1])

Qtd_Eleitores_pEstado = data.groupby('uf').sum().sort_values('total_eleitores',ascending=False)

w=0.2

values_x_fem = np.arange(len(Qtd_Eleitores_pEstado.index))
values_x_masc = [i+w for i in values_x_fem]
values_x_n_info = [i+w for i in values_x_masc]


values_y_fem = Qtd_Eleitores_pEstado['gen_feminino'].values
values_y_masc = Qtd_Eleitores_pEstado['gen_masculino'].values
values_y_n_info = Qtd_Eleitores_pEstado['gen_nao_informado'].values


col_fem = ax.bar(values_x_fem,values_y_fem,w,label='Feminino')
col_masc = ax.bar(values_x_masc,values_y_masc,w,label='Masculino')
col_n_info = ax.bar(values_x_n_info,values_y_n_info,w,label='Nao Informado')


plt.title('Quantidade de eleitores por genero',size=20)
plt.xticks(values_x_fem+w/3,Qtd_Eleitores_pEstado.index)
ax.set_ylabel('Quantidade de Eleitores',size=15)
ax.set_xlabel('UF',size=15, labelpad=10)


autolabel(col_fem)
autolabel(col_masc)
autolabel(col_n_info)


plt.legend()
plt.show()


print("Observação Importante: A quantidade aprensentada no eixo Y (por Exemplo 3.5) está em notação científica, e ao topo no canto esquerdo diz por qual numero deve ser multiplicado o valor do eixo Y para encontrar o valor real. \n\nNo topo de cada barra encontra-se o valor exato da quantidade de eleitores")
```
