```python
import plotly.graph_objects as go
```

```python
data.user_location.value_counts().head(30)
```

```python
# fig = go.Figure(go.Bar(x = data.user_location.value_counts().head(30).index, y = data.user_location.value_counts().head(30).values))
# fig.show()
# fig = go.Figure(go.Bar(x = data.source.value_counts().head(10).index, y = data.source.value_counts().head(10).values))
# fig.show()

fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = data.user_location.value_counts().head(30).index,
        y = data.user_location.value_counts().head(30).values,
        name = "Line"
    ))

fig.add_trace(
    go.Bar(
        x = data.user_location.value_counts().head(30).index,
        y = data.user_location.value_counts().head(30).values,
        name = "Bar"
    ))

fig.show()
```

```python
data.date = data.date.apply(lambda x: x.split(" ")[0])
```

```python
data.date.value_counts().sort_index()
```

```python
data.groupby(['date']).count().user_name.sort_index()
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = data.date.value_counts().sort_index().index,
        y = data.date.value_counts().sort_index().values,
        name = "Line"
    ))

fig.add_trace(
    go.Bar(
        x = data.date.value_counts().sort_index().index,
        y = data.date.value_counts().sort_index().values,
        name = "Bar"
    ))

fig.show()
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x = data.groupby(['date']).count().user_name.sort_index().index,
        y = data.groupby(['date']).count().user_name.sort_index().values,
        name = "Line"
    ))

fig.add_trace(
    go.Bar(
        x = data.groupby(['date']).count().user_name.sort_index().index,
        y = data.groupby(['date']).count().user_name.sort_index().values,
        name = "Bar"
    ))

fig.show()
```
