### Data Preprocessing

Extract year and season number from Season Name adn drop Season Name column

```python
season_Split = lambda x: pd.Series(str(x).split(' '))
df[['Season_drop','Season','Year']] = df.SeasonName.apply(season_Split)
```

```python
df.drop(['SeasonName','Season_drop'], axis=1, inplace = True)
```

Merge Driver First and last name as 1 column and drop both columns

```python
df['DriverName'] = df['DriverFirstName'] + ' ' + df['DriverLastName']
```

```python
df.drop(['DriverFirstName','DriverLastName'], axis=1, inplace = True)
```
