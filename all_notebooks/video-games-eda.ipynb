### Exploratory Data Analysis for Video games dataset
#### Import libraries needed

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../input/video-games/Games.csv',encoding='cp1252')
```

```python
df.head()
```

```python
df.shape
```

```python
# put the data in data frame
df = pd.DataFrame(df)
```

```python
df.describe()
```

```python
df['Review'].describe()
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
# print all games that took more than 5 score
richgames = df.loc[df['Score'] >= 5]
richgames.head()
```

```python
plt.scatter(df['Score'].iloc[:20],df['GameName'].iloc[:20],color='r')
```

```python
# print all games that took less than 5 score
poorgames = df.loc[df['Score'] <= 5]
poorgames.head()
```

```python
fig = plt.figure(figsize=(9,50))
fig = plt.plot(df['Review'].iloc[:200],df['GameName'].iloc[:200])
```

```python
df.Review.unique()
plt.pie(df['Review'].value_counts(),labels=['Good', 'Early Access', 'Great', 'Superb', 'Fair', 'Terrible',
       'Mediocre', 'Bad', 'Poor', 'Essential', 'Abysmal'])
```

```python
plt.bar(df['Review'].iloc[:20],df['GameName'].iloc[:20],color='r')
```
