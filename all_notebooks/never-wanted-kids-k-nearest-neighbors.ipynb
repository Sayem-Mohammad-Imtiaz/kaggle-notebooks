#Code by Alexander Bader https://www.kaggle.com/alexanderbader/perry-knn/data

```python
#Code by Alexander Bader https://www.kaggle.com/alexanderbader/perry-knn/data
#prepare data:

df=df.dropna()

qq1=[]
for i in df.user_explanations:
    zw = [int(s) for s in i.split() if s.isdigit()]
    if zw==[]:
        qq1.append([0.])
    else:
        qq1.append(np.mean(zw))
qq2=[]
for i in df.user_apparent_gender:
    zw = [int(s) for s in i.split() if s.isdigit()]
    if zw==[]:
        qq2.append([0.])
    else:
        qq2.append(np.mean(zw))
    

df.loc[:,"user_explanations"]=np.hstack(qq1).astype(int)
df.loc[:,"user_apparent_gender"]=np.hstack(qq2).astype(int)
```

```python
#Code by Alexander Bader https://www.kaggle.com/alexanderbader/perry-knn/data

#run knn-model, set 'n_neighbors' to desired no. of neighbors:

from sklearn.neighbors import NearestNeighbors

X = df[["number_upvotes","user_apparent_gender"]].values

nbrs = NearestNeighbors(n_neighbors=10, algorithm='ball_tree').fit(X)
nn_matrix = nbrs.kneighbors_graph(X).toarray()
```
