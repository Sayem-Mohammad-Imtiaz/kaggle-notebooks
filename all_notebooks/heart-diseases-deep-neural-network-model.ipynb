```python
dataset = pd.read_csv("/kaggle/input/heart-disease-dataset-uci/HeartDiseaseTrain-Test.csv")
```

```python
dataset.head()
```

```python
#This code changes categorical variables into different columns as different features. One hot encoding. Dummy variable trap is avoided
#Sex
liste_Male = []
for i in dataset["sex"]:
    if i == "Male":
        liste_Male.append(1)
    else:
        liste_Male.append(0)
        
#chest pain types
liste_Typical_angina = []
liste_Atypical_angina = []
liste_Non_anginal_pain = []
for i in dataset["chest_pain_type"]:
    if i == "Typical angina":
        liste_Typical_angina.append(1)
        liste_Atypical_angina.append(0)
        liste_Non_anginal_pain.append(0)
    elif i == "Atypical angina":
        liste_Typical_angina.append(0)
        liste_Atypical_angina.append(1)
        liste_Non_anginal_pain.append(0)
        
    elif i == "Non-anginal pain":
        liste_Typical_angina.append(0)
        liste_Atypical_angina.append(0)
        liste_Non_anginal_pain.append(1)
        
    else:
        liste_Typical_angina.append(0)
        liste_Atypical_angina.append(0)
        liste_Non_anginal_pain.append(0)
        
#fasting blood sugar
liste_sugar_higher = []
for i in dataset["fasting_blood_sugar"]:
    if i == "Greater than 120 mg/ml":
        liste_sugar_higher.append(1)
    else:
        liste_sugar_higher.append(0)
        
        
#ECG values
liste_ST_T_wave_abnormality = []
liste_Normal_wave = []
for i in dataset["rest_ecg"]:
    if i == "ST-T wave abnormality":
        liste_ST_T_wave_abnormality.append(1)
        liste_Normal_wave.append(0)
    elif i == "Normal":
        liste_ST_T_wave_abnormality.append(0)
        liste_Normal_wave.append(1)
    else:
        liste_ST_T_wave_abnormality.append(0)
        liste_Normal_wave.append(0)


liste_exercised_induced_angina = []
for i in dataset["exercise_induced_angina"]:
    if i == "Yes":
        liste_exercised_induced_angina.append(1)
    else:
        liste_exercised_induced_angina.append(0)
        
#slope
liste_Flat= []
liste_Upsloping=[]
for i in dataset["slope"]:
    if i == "Flat":
        liste_Flat.append(1)
        liste_Upsloping.append(0)
    elif i == "Upsloping":
        liste_Flat.append(0)
        liste_Upsloping.append(1)
    else:
        liste_Flat.append(0)
        liste_Upsloping.append(0)
#Vessels flourosopy    
for i in range(len(dataset["vessels_colored_by_flourosopy"])):
    if dataset["vessels_colored_by_flourosopy"].iloc[i] == "Zero":
        dataset["vessels_colored_by_flourosopy"].iloc[i] = 0
    
    elif dataset["vessels_colored_by_flourosopy"].iloc[i] == "One":
        dataset["vessels_colored_by_flourosopy"].iloc[i] = 1
        
    elif dataset["vessels_colored_by_flourosopy"].iloc[i] == "Two":
        dataset["vessels_colored_by_flourosopy"].iloc[i] = 2
        
    elif dataset["vessels_colored_by_flourosopy"].iloc[i] == "Three":
        dataset["vessels_colored_by_flourosopy"].iloc[i] = 3

    elif dataset["vessels_colored_by_flourosopy"].iloc[i]  == "Four":
        dataset["vessels_colored_by_flourosopy"].iloc[i] = 4
        
        
#thalassemia
liste_No_defect = []
liste_Normal_defect = []
liste_Fixed_defect = []
for i in dataset["thalassemia"]:
    if i == "No":
        liste_No_defect.append(1)
        liste_Normal_defect.append(0)
        liste_Fixed_defect.append(0)
        
    elif i == "Normal":
        liste_No_defect.append(0)
        liste_Normal_defect.append(1)
        liste_Fixed_defect.append(0)
        
    elif i == "Fixed Defect":
        liste_No_defect.append(0)
        liste_Normal_defect.append(0)
        liste_Fixed_defect.append(1)
        
    else:
        liste_No_defect.append(0)
        liste_Normal_defect.append(0)
        liste_Fixed_defect.append(0)
        
```

```python
final_liste = [liste_Male,liste_Typical_angina,liste_Atypical_angina,liste_Non_anginal_pain,liste_sugar_higher,liste_ST_T_wave_abnormality,liste_Normal_wave,liste_exercised_induced_angina,liste_Flat,liste_Upsloping,liste_No_defect,liste_Normal_defect,liste_Fixed_defect]

final_liste_names = ["liste_Male","liste_Typical_angina","liste_Atypical_angina","liste_Non_anginal_pain","liste_sugar_higher","liste_ST_T_wave_abnormality","liste_Normal_wave","liste_exercised_induced_angina","liste_Flat","liste_Upsloping","liste_No_defect","liste_Normal_defect","liste_Fixed_defect"]
```

```python
#Here we added our categorical variables as columns into the dataframe
for i in range(len(final_liste_names)):
    dataset[final_liste_names[i]] = list(final_liste[i])
    print(final_liste_names[i])
```

```python
liste_target = list(dataset["target"])
```

```python
#We get rid of old categorical columns since we have encoded them.
dataset = dataset.drop(["sex","chest_pain_type","fasting_blood_sugar","rest_ecg","exercise_induced_angina","slope","thalassemia","target"],axis = 1)
```

```python
# Here we add our target column again to make slicing easier.
dataset["target"] = list(liste_target)
```

```python
dataset.head()
```

```python
#We create the numpy values
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
#Splitting data into train and test sets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 5)
```

```python
#We scale the values to between 0 and 1
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Here we create a deep neural network with two hidden layers.
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=100,activation = "relu"))
ann.add(tf.keras.layers.Dense(units = 100,activation = "relu"))
ann.add(tf.keras.layers.Dense(units = 1,activation = "sigmoid"))
ann.compile(optimizer = "adam",loss = "binary_crossentropy",metrics = ["accuracy"])
```

```python
#We train the neural network here with 100 epoch with batch size of 32 for the backpropagation procedure.
ann.fit(X_train,y_train,batch_size = 32,epochs = 150 )
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
#We can check the success of the deep neural network. It predicts all the cases correctly in the test set. We can see from the confusion matrix and accuracy score.
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
