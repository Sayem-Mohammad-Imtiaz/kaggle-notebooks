```python
import pandas as pd
import numba as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
data=pd.read_csv("../input/covid19-in-indonesia/Covid 19 in Indonesia.csv")
```

```python
data.info()
```

```python
data['date']=pd.to_datetime(data['date'])
```

```python
fig,axis=plt.subplots(2,2,figsize=(20,13))
plt.style.use("fivethirtyeight")
sns.lineplot(x='date', y='jumlah_meninggal',data=data,color='red', ax=axis[0,0])
sns.lineplot(x='date', y='jumlah_sembuh',data=data,color='black', ax=axis[0,1])
sns.lineplot(x='date', y='jumlah_positif',data=data,color='green', ax=axis[1,0])
sns.lineplot(x='date', y='jumlah_dirawat',data=data, color='yellow',ax=axis[1,1])
```

```python
fig,axis=plt.subplots(2,2,figsize=(20,13))
plt.style.use("ggplot")
sns.lineplot(x='date', y='jumlah_positif_kum',data=data,color='brown', ax=axis[0,0])
sns.lineplot(x='date', y='jumlah_sembuh_kum',data=data,color='blue', ax=axis[0,1])
sns.lineplot(x='date', y='jumlah_meninggal_kum',data=data,color='Pink', ax=axis[1,0])
sns.lineplot(x='date', y='jumlah_dirawat_kum',data=data, color='grey',ax=axis[1,1])
```
