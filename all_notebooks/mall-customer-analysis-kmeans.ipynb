```python
x = df_mall.iloc[:, [3, 4]].values 
gender = {'Male': 1,'Female': 2}
df_mall.Gender = [gender[item] for item in df_mall.Gender]
df_mall.drop('CustomerID', axis=1, inplace=True)
```

```python
# Elbow method to finding the No of clusters for the dataset.
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(8,8))
wcss_list = []

for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=100)
    kmeans.fit(df_mall)
    wcss_list.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss_list)
plt.xlabel('No.of Clusters')
plt.ylabel('wcss list')
plt.show()
```

* No of cluster is 4 for the mall dataset using the elbow method

```python
# Cluster Prediction using KMeans

kmeans = KMeans(n_clusters=5, init='k-means++', random_state=100)
y_pred = kmeans.fit_predict(df_mall)
```

```python
len(y_pred)
```

```python
df_mall['Predicted']  = y_pred
```

# Conclusion

* Cluster1 shows the customers with average salary and average spending so we can categorize these customers as **average**.
* Cluster2 shows the customer has a high income but low spending, so we can categorize them as **careful**.
* Cluster3 shows the customers with high income and high spending so they can be categorized as **target**, and these customers can be the most profitable customers for the mall owner.
* Cluster4 shows the low income and also low spending so they can be categorized as **sensible**.
* Cluster5 shows the customers with low income with very high spending so they can be categorized as **careless**.
