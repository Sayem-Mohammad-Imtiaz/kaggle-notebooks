# Veri Bilimi Notları (Türkçe)

Burada genel olarak veri bilimiyle ilgili notlarım ve örneklerim bulunmaktadır. Piyasadaki eksik bilgilerden ve yapancı bilgilerden ötürü hepsini tek bir seferde bulabileceğiniz notlardan oluşmaktadır. Umarım işinize yarar.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns  # visualization tool
```

```python
data = pd.read_csv('../input/pokemon-challenge/pokemon.csv')
```

```python
data.head(5)
```

# Line Plot

Görüldüğü gibi burada çizgisel bir grafik elde ettik ve burada pokemon sayılarını zaman olarak da düşünebiliriz. Zamana göre cismin aldığı değerler olabilir burada kullanabiliriz.

```python
data.Speed.plot(kind = 'line', color = 'g',label = 'Speed',linewidth=1,alpha = 0.5,grid = True,linestyle = '-') #alpha demek saydamlık 
data.Attack.plot(kind = 'line', color = 'r', label = 'Attack', linewidth = 1 , alpha = 0.5, grid = True, linestyle = ':')
plt.legend(loc='upper right')
plt.xlabel('Number of Pokemons')
plt.ylabel('Speed and Attack Levels')
plt.title('Line Plot')
plt.show()
```

# **Scatter Plot Kullanımı**

Bununlar çeşitli şeyleri kıyaslayacağız korelasyon da denilebilir yani correlation. 

```python
data.plot(kind = 'scatter', x = 'Attack', y = 'Defense', alpha = 0.5, color = 'red')
plt.title('Attack and Defense Scatter Plot')
plt.show()
```

**Farklı bir biçimde scatter plot kullanımı**

Bu aşağıdaki şekilde de yukarıdakinin aynısını yapabiliyoruz hiç bir şey değişmiyor. 

```python
plt.scatter(data.Attack,data.Defense, color = 'red', alpha = 0.5)
plt.show()
```

# **Historgram Plot**

adı üstünde histogram yöntemiyle gösteriyor

```python
data.Speed.plot(kind = 'hist', bins = 50, figsize = (15,15)) #bins is number of bar in figure
plt.xlabel('The Values of Speed')
plt.show()
```

# **While and For Loops**

```python
i = 0
while i != 5:
    print("i = ", i)
    i += 1

print(i, ", 5e eşittir.")
```

```python
 lis = [1,2,3,4,5]

for i in lis:
    print(i)
print('')

dictionary = {'spain':'madrid','france':'paris'}
for key, value in dictionary.items():
    print(key, " : ", value)
print('')

for index, value  in data[['Attack']][0:1].iterrows():
    print(index, " : ", value)
```
