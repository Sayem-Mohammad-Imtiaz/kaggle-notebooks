```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import precision_score, recall_score, f1_score
from sklearn.metrics import plot_roc_curve
```

```python
df = pd.read_csv('../input/heart-disease-uci/heart.csv')
df.head()
```

Splitting the data 

```python
x = df.drop('target',axis = 1)
y = df['target']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

Model Building

## RandomForestClassifier

```python
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_preds = model.predict(x_test)
```

```python
print(classification_report(y_test,y_preds))
```

```python
conf = confusion_matrix(y_test,y_preds)
conf
```

```python
f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(conf,
            annot=True,
            linewidths=0.5,
            linecolor="red",
            fmt=".0f",
            ax=ax)
plt.xlabel("Predicted Values")
plt.ylabel("True Values")
plt.show();
```

## KNeighborsClassifier

```python
model1 = KNeighborsClassifier()
model1.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
kn_preds = model1.predict(x_test)
```

```python
print(classification_report(y_test,kn_preds))
```

```python
conf_mat = confusion_matrix(y_test,kn_preds)
conf_mat
```

```python
f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(conf_mat,
            annot=True,
            linewidths=0.5,
            linecolor="red",
            fmt=".0f",
            ax=ax)
plt.xlabel("Predicted Values")
plt.ylabel("True Values")
plt.show();
```

## LogisticRegression

```python
model2 = LogisticRegression()
model2.fit(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```

```python
log_preds = model2.predict(x_test)
```

```python
print(classification_report(y_test,log_preds))
```

```python
mat = confusion_matrix(y_test,log_preds)
mat
```

```python
f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(mat,
            annot=True,
            linewidths=0.5,
            linecolor="red",
            fmt=".0f",
            ax=ax)
plt.xlabel("Predicted Values")
plt.ylabel("True Values")
plt.show();
```

## SVC

```python
model3 = SVC()
model3.fit(x_train,y_train)
```

```python
model3.score(x_test,y_test)
```

```python
svc_preds = model3.predict(x_test)
```

```python
print(classification_report(y_test,svc_preds))
```

```python
svc_mat = confusion_matrix(y_test,svc_preds)
svc_mat
```

```python
f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(svc_mat,
            annot=True,
            linewidths=0.5,
            linecolor="red",
            fmt=".0f",
            ax=ax)
plt.xlabel("Predicted Values")
plt.ylabel("True Values")
plt.show();
```

Plotting the scores of all model

```python
scores = pd.DataFrame({'RandomForest': model.score(x_test,y_test),
                       'KNeighborsClassifier': model1.score(x_test,y_test),
                       'LogisticRegression': model2.score(x_test,y_test),
                       'SVC': model3.score(x_test,y_test)},
                        index = [0])
```

```python
scores.T.plot(kind = 'bar',
              figsize = (10,10))
plt.title('Scores of all Model')
plt.xlabel('Model Name')
plt.ylabel('Scores');
              
                      
```
