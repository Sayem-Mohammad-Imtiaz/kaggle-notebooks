****Amazon top 50 bestselling books****

*seaborn visualization*


Read data and quick analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

```python
file_path = "../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv"
book = pd.read_csv(file_path, index_col="Name")
book.head()
```

Check for missing data.

```python
book.isnull().sum()
```

**Analysis by genre of book**

The top selling books by genre. Non fiction books are more popular throughout  the users. However the non fiction books costs more the fiction.


```python
genre_price = book.groupby(["Genre"])["Price"].mean()
genre_book_count = book["Genre"].value_counts()
fig1 = plt.figure(figsize=(8, 8))
plt.pie(genre_book_count, explode=(0, 0), labels=["Non Fiction", "Fiction"], colors = ["orange", "green"], autopct="%1.1f%%",
     startangle=90, textprops={"fontsize": 16, "weight": "bold"})
plt.show()
```

**Relationship between Price and Users Rating**

The distribution of books by users rating and price. The highest numbers of books received rating in range of 4.6 - 4.8 points. The price of that books are average.

```python
fig2 = plt.figure(figsize=(10,10))
sns.scatterplot(x=book["User Rating"], y=book["Price"], data=book, size="Price", hue="Price", sizes=(100, 200))
plt.ylabel("Price",fontsize=16)
plt.xlabel("User Rating", fontsize=16)
plt.show()
```

```python
highest_rating = book.loc[(book["User Rating"] >= 4.6)
                           & (book["User Rating"] < 4.81)]
```

```python
fig3 = plt.figure(figsize=(15, 10))
palettes = sns.color_palette("Set2")
ax = sns.boxplot(x=highest_rating["User Rating"], y=highest_rating["Price"], data=book, showfliers = False, palette=palettes, width=0.4, whis=10, linewidth=2.5)
ax = sns.swarmplot(x=highest_rating["User Rating"], y=highest_rating["Price"], data=book, color=".25")
#ax = sns.stripplot(x=highest_rating["User Rating"], y=highest_rating["Price"], data=book, color=".25", orient="v", )
plt.xlabel("The most common users rating", fontsize=16)
plt.ylabel("Price",fontsize=16)
ax.yaxis.set_major_locator(ticker.MultipleLocator(10))

plt.show()
```

**Relationship between Price and Reviews**

More pricey books get less reviews. However the most reviews received books with average price. 


```python
fig4 = plt.figure(figsize=(10, 10))
sns.set_style("whitegrid")
palettes = sns.color_palette("crest", as_cmap=True)
sns.scatterplot(y=book["Price"], x=book["Reviews"], data=book, hue="Reviews", size="Reviews", palette=palettes)
plt.xlabel = book["Reviews"]
plt.ylabel = book["Price"]
plt.xlim(-1,)
plt.ylim(-1,)
plt.show()
```

**Average Prices in a year**

In the 2015 year the avarage price reach aproximately 10$. However the price of the bestsellers until 2015 was higher. After a sudden rise in book prices in 2016, book prices declined in the following years. 

```python
fig5, ax1 = plt.subplots(figsize=(15,10))
sns.set_style("white")
ax2 = ax1.twinx()
book["Mean_price"] = book.groupby(["Year"])["Price"].transform("mean")
sns.lineplot(data=book, x=book["Year"], y=book["Mean_price"], ax=ax1)
sns.barplot(y=book["Price"], x=book["Year"], data=book, capsize=.2, ax=ax2)
plt.show()
```

**Average numbers of reviews over the years**

The numbers of reviews rises throughout the years more or less exponentially.

```python
fig6 = plt.figure(figsize=(15,10))
sns.barplot(y=book["Reviews"], x=book["Year"], data=book, palette="Blues_d")
plt.show()
```

**The authors of the most rated bestsellers**

The top 10 authors with most highest rated books. 
Jeff Kinney has the most books which got at the list of the 50 Bestsellers.

```python
temp1 = book.groupby("Author").count().sort_values("User Rating", ascending=False).head(10)
temp1["User Rating"]
```

The relationship between user rating and reviews of the top 10 authors.

```python
fig7 = plt.figure(figsize=(10,10))
sns.set_style("whitegrid")
temp2 = book.groupby("Author").mean().sort_values("Reviews", ascending=False).head(10)
sns.scatterplot(data=temp2, x="User Rating", y="Reviews", hue="Author", s=200)
plt.ylim(30000, 90000)
plt.show()
```

The relationship between user rating and books price of the top 10 authors.

```python
fig8 = plt.figure(figsize=(10,10))
sns.scatterplot(data=temp2, x="User Rating", y="Price", hue="Author", s=200)
plt.ylim(8, 40)
plt.show()
```
