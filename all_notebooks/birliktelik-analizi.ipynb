**Apriori Algoritması :** Kullanıcı alışverişlerinden ürün birlikteliklerini ortaya çıkarmak için kullanılan yöntemtir. Belirlenecek bir eşik değere göre satın alınan ürünlerin birlikteliklerini görme imkanı sağlar.


- ADIM 1: Her bir ürünün support'unu (frekans) hesaplama
- ADIM 2: Supporta göre eleme yap
- ADIM 3: Yeni liste ve supportu oluştur
- ADIM 4: Supporta göre eleme yap 
- ADIM 5: Yeni liste ve supportu oluştur
- ADIM 6: Supporta göre eleme yap
- ADIM 7: Final tablosunu oluşturup yorumla

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_csv('../input/grocerystoreds/GroceryStoreDataSet.csv', names=['products'],header=None)
df.head()
```

```python
df.shape
```

```python
df
```

```python
df.columns 
```

```python
df.values
```

```python
data = list(df["products"].apply(lambda x:x.split(',')))
data 
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
# one-hot encoding dönüşüm işlemi 
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data,columns=te.columns_)
df
```

```python
freq_items = apriori(df, min_support = 0.2, use_colnames = True, verbose = 1)
```

```python
freq_items
```

```python
freq_items.sort_values(by="support", ascending = False)
```

```python
new_df = association_rules(freq_items, metric="confidence", min_threshold = 0.25)
new_df.sort_values(by ="confidence", ascending = False)
```

Threshold değerini 0.25 değeri verilip, "confidence" değerine göre azalan şekilde sıralandı. Confidence değeri 0.80 olan MILK-BREAD ikilisine bakıldığında birlikte satılma değerleri %0.2. Süt alan birinin sepetinde ekmeğide görme ihtimalimiz ise %0.8'dir.  Confidence değerine göre azalan şekilde düzenlediğimiz veri setimize baktığımızda kahvenin alınma değeri %0.4 , Cornflakes'in ise tek başına alınma ihtimali %0.30, birlikte satılma değerleri %0.2 , Kahve alan birinin Cornflakes alma ihtimali ise %0.50'dir. Son olarak bir ikiliye de bakalım ve değerlendirme ve satışların artması için markete değişiklik önerilerinde bulunalım. Normal olarak ekmeğin tek başına satılma değeri ürünler arasında en fazla değere sahip üründür. Çayın ise tek başına %0.35 değerine sahiptir. BREAD-TEA ikilisini birlikte satılma oranı %0.2, ekmek alan birinin sepetinde çay da görme durumunun değeri ise %0.30'dur. 

Burada kahve ve alan birinin cornflakes gibi atıştırmalık tercih ettiğini gördüğümüz için kahve reyonuna yakın olacak şekilde atıştırmalık yiyecekleri (çikolata, bisküvi gibi) yerleştirilmesini öneririm.
