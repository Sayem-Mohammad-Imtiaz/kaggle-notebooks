```python
sns.countplot('Partner',data=df,hue='Churn')
# So the customers who have partners will have multiple plans and that's why would be less likely to get Churn Out.
```

```python
sns.countplot('Dependents',data=df,hue='Churn')
# So the customers who have dependents will also have multiple plans and that's why would be less likely to get Churn Out.
```

```python
#The Old the Customer is the more Loyal he is to the Company - Deep Gojariya(2021)
sns.set(style="whitegrid")
sns.violinplot('Churn','tenure',data=df,orient="v")
#Here we can clearly see that most of the people who churned out were customers for less than 5 months
```
