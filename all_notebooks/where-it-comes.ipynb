In this Notebook we explain how we get this dataset.

The Magellan mission to Venus (https://nssdc.gsfc.nasa.gov/imgcat/html/object_page/mgn_f20s003_1.html) to scan the ground via radar images was the inspired dataset that we choose to create this. It was composed of radar images with the coordinate of possible volcanoes.
![image1](https://i.imgur.com/xawRxI9.png)

We takes this images and the coordinates and mark them.
![image2](https://i.imgur.com/2l8jSeY.png)

After that we crop and create the dataset that we present here in Kaggle.

![image3](https://i.imgur.com/ATRPezn.png)
