# Load Dataset

```python
df=pd.read_csv('/kaggle/input/pima-indians-diabetes-database/diabetes.csv')
```

# <h3> Checking Correlation Between Each Feature

```python
plt.figure(figsize=(8,5))
sns.heatmap(df.corr(),annot=True)
```

```python
X = df.drop(['Outcome'], axis=1)
y = df['Outcome']
```

# Split the data for tranning and testing

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=18) 
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
```

# SVC Model

```python
svc = SVC()
params = {'kernel':('linear', 'rbf', 'poly', 'sigmoid'), 'gamma':[1, 0.1, 0.01, 0.001, 0.0001],'C':[1,10]}
model = GridSearchCV(svc, params)
model.fit(X_train, y_train)
print(model.best_params_)
```

GridSearchCV() is a library function in the sklearn's model_sekection package. It is used to select the optiomal hyperameters for the model. 

# <h3> Test Model

```python
X_test = scaler.transform(X_test)
y_pred = model.predict(X_test)
```

# Acurracy and Confusion Matrix of the model

```python
accuracy = accuracy_score(y_test, y_pred)
con_matrix = confusion_matrix(y_test, y_pred)

print(f"Acurracy: {accuracy}")
print(f"Con_Matrix: {con_matrix}")
```
