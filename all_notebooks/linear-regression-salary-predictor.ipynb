```python
# 1. Extract your input and output
# 2. Divide your data into training and testing set
# 3. Train your model
# 4. Do Prediction
```

```python
df['YearsExperience'].values
```

```python
X = df['YearsExperience'].values.reshape(30,1)
```

```python
X.shape
```

```python
Y = df['Salary'].values
```

```python
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test  = train_test_split(X,Y, test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
reg.fit(X_train,Y_train)
```
