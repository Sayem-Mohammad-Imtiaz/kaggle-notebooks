# **Grafik çizimleri**

Öncelikle veri kümesinde bulunan her bir özelliğin birbiriyle olan ilişkisini görmek için korelasyon değerlerine bakalım:

```python
f,ax = plt.subplots(figsize=(13,13))
sns.heatmap(data.corr(),annot=True, linewidth=.5,fmt='.2f',ax=ax)
plt.show()
```

Korelasyon tablosuna bakarak, sigara tüketimi ve cinsiyet arasında bir ilişki olduğunu,zaman ve ölüm arasında bir ilişki olmadığı söylenebilir.

```python
plt.scatter(data.smoking,data.serum_creatinine,color="blue")
plt.show()
```

```python
#Kreatin kinaz ile yaş arasındaki korelasyonun incelenmesi
plt.scatter(data.age, data.creatinine_phosphokinase, s=10 ) #s used for marker size
plt.title("Yaş ve Kreatin Kinaz ilişkisi")
plt.xlabel("Yaş")
plt.ylabel("Kreatin Kinaz")
plt.show()
```

```python
#Kalp yetmezliği ile yaş arasındaki korelasyonun incelenmesi
plt.scatter(data.age, data.ejection_fraction, s= 10) 
plt.title("Kalp yetmezliği ve Yaş ilişkisi")
plt.xlabel("Yaş")
plt.ylabel("Kalp Yetmezliği")
plt.show()
```

```python
import plotly.express as px

ds = data['high_blood_pressure'].value_counts().reset_index()
ds.columns = ['high_blood_pressure', 'count']
fig = px.bar(ds, x='high_blood_pressure', y="count", title='Count of people with High-blood pressure', width=400)
fig.show()
```

```python
corr = data.corr()
fig = px.imshow(corr) 
fig.show()
```
