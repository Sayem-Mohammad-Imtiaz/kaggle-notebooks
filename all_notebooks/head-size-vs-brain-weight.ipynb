## Evaluate Model By Basic Mathematics

### For y = mx + c  
### Lets find Intercept(c) and Slope(m)

```python
# finding mean of X & y
mean_X = np.mean(X) 
mean_y = np.mean(y)

# Calculating the value of m & c using Formula
num = 0
den = 0

for i in range(len(X)):
    num = num + (X[i] - mean_X ) * (y[i] - mean_y)
    den = den + (X[i] - mean_X ) **2
    
m = num / den
c = mean_y - ( m * mean_X )

print('Slope = ',m)
print('intercept = ',c)
```

### Plotting Actual Points & Regression Line

```python
max_x = np.max(X) + 100
min_x = np.min(X) - 100

#Calculating line values of x & y
X1 = np.linspace(min_x, max_x, 1000)
y1 = c + m*X1
```

```python
plt.figure(figsize=(10,10))
# Plot Line
plt.plot(X1, y1,color='r', label = 'Regression Line')

# Plotting Actual Points
plt.scatter(X, y , color = 'k', label = 'Actual Points')

plt.xlabel('HEAD Size')
plt.ylabel('BRAIN Weight')
plt.legend()
plt.show()
```

### Predict the Value of y

```python
y_pred = []
for i in range(len(X)):
    y_new = m*X[i] + c
    y_pred.append(y_new)
print(y_pred)
```

### Model accuracy
find Model accuracy  by using **Mean squared error, Root mean squared error** and **R Squared Method**.

```python
ss_t = 0
ss_r =0
for i in range(len(X)):
    ss_r += (y[i] - y_pred[i]) ** 2
    ss_t += (y[i] - mean_y) ** 2

mse = ss_r/len(X)
r_mse = np.sqrt(mse)    
r_sqr = 1 - (ss_r/ ss_t )
```

```python
print('mean squared error: ',mse)
print('Root mean squared error: ',r_mse)
print('R_Squared: ',r_sqr)
```

## Evaluate Model BY LINEAR REGRESSION

### Reshape the input data into 2D array

```python
X = X.reshape(len(X),1)
```

### Creating Model

```python
reg = LinearRegression()
```

### Fitting Training Data

```python
 reg = reg.fit(X,y)
```

### Y Prediction

```python
y_pred = reg.predict(X)
```

### Calculate RMSE and R Square

```python
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
r2_score = reg.score(X,y)
```

```python
print('mean squared error: ',mse)
print('Root mean squared error: ',rmse)
print('R_Squared: ',r2_score)
```
