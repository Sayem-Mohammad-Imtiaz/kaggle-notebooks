**Montgomery–Åsberg Depresyon Derecelendirme Ölçeği**

MADRS, antidepresan etkinlik denemelerinde en sık kullanılan sonuç ölçüsüdür ve son on yılda artan sıklıkta kullanılmıştır. Bu ölçeğin temel amacı, depresif semptomolojiyi, özellikle antidepresanlarla tedavi edilen hastalardaki değişiklikleri değerlendirmektir.

HAM-D17 somatik belirtilere yönelik maddeler içerirken, MADRS depresyonun psikolojik belirtilerine (örneğin, üzüntü, gerilim ve karamsar düşünceler) odaklanır.Bu ölçek klinisyen tarafından derecelendirilmiştir ve 10 maddeden oluşmaktadır; her madde 0-6 arasında derecelendirilir, bu da maksimum toplam 60 puanla sonuçlanır ve daha yüksek puanlar daha fazla depresif semptomolojinin göstergesidir.

MADRS puanlama yönergeleri, 0 ile 6 arasında değişen toplam puanın hastanın normal aralıkta olduğunu (depresyon olmadığını), 7 ile 19 arasında değişen bir puanın "hafif depresyon", 20 ile 34 arasında "orta depresyon" olduğunu gösterir. 35 ve üzeri puan “ağır depresyon”u, toplam 60 puan ve üzeri ise “çok şiddetli depresyon”u gösterir
