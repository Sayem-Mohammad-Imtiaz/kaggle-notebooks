```python
import sklearn
from sklearn.preprocessing import LabelEncoder
```

```python
data1=data1.apply(LabelEncoder().fit_transform)
```

```python
data1.info()
```

```python
_=klib.corr_plot(data=data1, target='class')
```

```python
Y=data1['class']
X=data1.drop(['class'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train,Y_test=train_test_split(X,Y, test_size=0.2, random_state=10)
```

```python
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier()
model=RF.fit(X_train, Y_train)
Y_pred=model.predict(X_test)
Y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print('Accuracy value is', accuracy_score(Y_pred,Y_test))
```

```python
import seaborn as sns
_=sns.heatmap(confusion_matrix(Y_pred, Y_test), annot=True, fmt='g', cmap='Blues')

```
