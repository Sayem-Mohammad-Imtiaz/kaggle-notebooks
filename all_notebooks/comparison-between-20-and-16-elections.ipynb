```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import os
import seaborn as sns
import statistics
from scipy import stats
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
plt.style.use('dark_background')
```

```python
df=pd.read_csv('/kaggle/input/2020-general-election-polls/county_statistics.csv') #importing our data
#df = df.dropna(axis=0)
```

Let's start with some general data comparison between 2016 and 2020

```python
difference_votes = (df["total_votes20"] - df["total_votes16"])/max(df["total_votes20"] - df["total_votes16"])*100
difference_votes = difference_votes.dropna(axis=0)
difference_votes.plot.hist()
plt.yscale('log')
plt.title("Number of states for increase in number of voter (%), yscale is log")
plt.xlabel("% difference in voters between 2016 and 2020")
plt.ylabel("Number of counties (logscale)")
plt.show()
```

Some states record a decrease in voters but most states record an increase or stable number of voters.

```python
diff = df["total_votes20"] - df["total_votes16"]
diff = diff.dropna(axis=0)
sum_total = sum(diff)
print("Absolute difference in voters in the country: {}".format(int(sum_total)))
```

This confirms a global increase.

```python
tr = df.loc[df['votes20_Donald_Trump'] > df['votes20_Joe_Biden']]
difference_votes = (tr["total_votes20"] - tr["total_votes16"])/max(tr["total_votes20"] - tr["total_votes16"])*100
difference_votes = difference_votes.dropna(axis=0)
difference_votes.plot.hist(color="red")
plt.yscale('log')
plt.title("Number of states for increase in number of voter among couties DT won (%), yscale is log")
plt.xlabel("% difference in voters between 2016 and 2020")
plt.ylabel("Number of counties (logscale)")
plt.show()
diff = tr["total_votes20"] - tr["total_votes16"]
diff = diff.dropna(axis=0)
sum_total = sum(diff)
print("Absolute difference in voters in counties that voted DT in 2020: {}".format(int(sum_total)))
```

```python
bi = df.loc[df['votes20_Donald_Trump'] < df['votes20_Joe_Biden']]
difference_votes = (bi["total_votes20"] - bi["total_votes16"])/max(bi["total_votes20"] - bi["total_votes16"])*100
difference_votes = difference_votes.dropna(axis=0)
difference_votes.plot.hist(color="blue")
plt.yscale('log')
plt.title("Number of states for increase in number of voter among couties JB won (%), yscale is log")
plt.xlabel("% difference in voters between 2016 and 2020")
plt.ylabel("Number of counties (logscale)")
plt.show()
diff = bi["total_votes20"] - bi["total_votes16"]
diff = diff.dropna(axis=0)
sum_total = sum(diff)
print("Absolute difference in voters in counties that voted JB in 2020: {}".format(int(sum_total)))
```

```python
hc_jb = df.loc[(df['votes20_Donald_Trump'] < df['votes20_Joe_Biden'])]
hc_dt = df.loc[(df['votes20_Donald_Trump'] > df['votes20_Joe_Biden'])]
plt.bar(["JB", "DT"], [len(hc_jb.index), len(hc_dt.index)],color=["blue","red"])
[len(hc_jb.index), len(hc_dt.index)]
plt.title("Candidate voted by counties")
plt.xlabel("Candidate")
plt.ylabel("Number of counties")
plt.show()
```

Many more counties voted DT in 2020.

```python
hc_jb = df.loc[(df['votes20_Donald_Trump'] < df['votes20_Joe_Biden']) & (df['votes16_Donald_Trump'] < df['votes16_Hillary_Clinton'])]
hc_dt = df.loc[(df['votes20_Donald_Trump'] > df['votes20_Joe_Biden']) & (df['votes16_Donald_Trump'] < df['votes16_Hillary_Clinton'])]
plt.bar(["JB", "DT"], [len(hc_jb.index), len(hc_dt.index)],color=["blue","red"])
[len(hc_jb.index), len(hc_dt.index)]
plt.title("Candidate voted by counties that voted HC in 2016")
plt.xlabel("Candidate")
plt.ylabel("Number of counties")
plt.show()
print("Number of couties that shifted to from HC to DT: {}".format(len(hc_dt.index)))
```

```python
dt_jb = df.loc[(df['votes20_Donald_Trump'] < df['votes20_Joe_Biden']) & (df['votes16_Donald_Trump'] > df['votes16_Hillary_Clinton'])]
dt_dt = df.loc[(df['votes20_Donald_Trump'] > df['votes20_Joe_Biden']) & (df['votes16_Donald_Trump'] > df['votes16_Hillary_Clinton'])]
plt.bar(["JB", "DT"], [len(dt_jb.index), len(dt_dt.index)],color=["blue","red"])
[len(dt_jb.index), len(dt_dt.index)]
plt.title("Candidate voted by counties that voted DT in 2016")
plt.xlabel("Candidate")
plt.ylabel("Number of counties")
plt.show()
print("Number of couties that shifted to from DT to JB: {}".format(len(dt_jb.index)))
```

```python
plt.figure(figsize=(12,8))
hm = sns.heatmap(df[["state", "votes16_Donald_Trump", "votes16_Hillary_Clinton","votes20_Donald_Trump", "votes20_Joe_Biden"]].corr(), annot = True) #overall correlation between the various columns present in our data
plt.title('Overall relation between columns of the Dataset', fontsize = 20)
plt.show()
```

We see high correlation betweend all these variables. This show few changes between elections
