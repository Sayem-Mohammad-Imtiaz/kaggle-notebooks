```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
import re
from nltk.stem import PorterStemmer
import gensim
import gensim.corpora as corpora
```

```python
reviews = pd.read_csv("../input/wine-beer-and-liquor-reviews/wine reviews.csv")
reviews = reviews["reviews.text"].str.replace('[^\w\s]','')
porter_stemmer=PorterStemmer()
```

```python
stop_words = set(stopwords.words('english'))
```

```python
for index, value in reviews.items():
    reviews[index] = [porter_stemmer.stem(word=w) for w in nltk.word_tokenize(str(reviews[index]).lower()) if not w.lower() in stop_words]
    #reviews[index] = [w for w in nltk.word_tokenize(str(reviews[index]).lower()) if not w.lower() in stop_words]
    
series = []
for i, value in reviews.items():
    for word in value:
        series.append(str(word))
    
#print(series)
```

```python
from wordcloud import WordCloud
# Join the different processed titles together.
long_string = ','.join(series)
# Create a WordCloud object
wordcloud = WordCloud(background_color="white", max_words=50, contour_width=3, contour_color='steelblue')
# Generate a word cloud
wordcloud.generate(long_string)
# Visualize the word cloud
wordcloud.to_image()
```

```python
# Create Dictionary
id2word = corpora.Dictionary([series])
# Create Corpus
texts = [series]
# Term Document Frequency
corpus = [id2word.doc2bow(text) for text in texts]
#corpus
```

```python
from pprint import pprint
# Build LDA model
lda_model = gensim.models.LdaMulticore(corpus=corpus, id2word=id2word, num_topics=10)
# Print the Keyword in the 10 topics
topics_list = lda_model.print_topics()
doc_lda = lda_model[corpus]
print(topics_list[0])
```

```python
for i in topics_list:
    topics = str(i).split('+')
    #print(str(topics)+"\n")
    regex = re.compile('[^a-zA-Z]')
    topic_lists = []
    for topic in topics:
        topic = regex.sub('', topic)
        topic_lists.append(topic)
    print(topic_lists)
    long_string = ','.join(topic_lists)
    # Create a WordCloud object
    wordcloud = WordCloud(background_color="white", max_words=5, contour_width=1, contour_color='steelblue')
    # Generate a word cloud
    wordcloud.generate(long_string)
    # Visualize the word cloud
    wordcloud.to_image()
```
