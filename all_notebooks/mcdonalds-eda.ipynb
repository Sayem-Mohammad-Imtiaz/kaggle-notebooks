```python
mc = pd.read_csv('/kaggle/input/mcdonald-data-analysis/McD.csv')
```

```python
mc.head()
```

```python
mc1 = pd.DataFrame(mc.groupby('Category')['Item'].nunique()).sort_values(by= 'Item' , ascending = False)
```

```python
plt.figure(figsize=(20,10))
sns.countplot( mc['Category'] , order = mc['Category'].value_counts().index)
plt.xticks(rotation = 90);
plt.title('Food categories with different variety' , fontsize = 20)
plt.show()
```

### From above graph , food category with highest variety is 'coffee & Tea' and food category with lowest variety is ' salad '

```python
plt.figure(figsize=(10,25))
plt.xlabel('x-axis' , fontsize =20)


list = ['Calories' , 'Calories from Fat' , 'Total Fat' , 
        'Total Fat (% Daily Value)' , 'Saturated Fat' , 'Saturated Fat (% Daily Value)' , 'Trans Fat',
       'Cholesterol' , 'Cholesterol (% Daily Value)' , 'Sodium' , 'Sodium (% Daily Value)' , 'Carbohydrates',
       'Carbohydrates (% Daily Value)' ,'Dietary Fiber', 'Dietary Fiber (% Daily Value)' , 'Sugars' , 'Protein' ,'Vitamin A (% Daily Value)', 
       'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)' , 'Iron (% Daily Value)']
for i in range(0,21):
    plt.subplot(21,1,i+1)
    plt.subplots_adjust(hspace=1)
    sns.boxplot(mc[list[i]] , width = 2 , linewidth=1 , fliersize=5 , orient= 'h' )
  
```

```python
chol = pd.DataFrame(mc.groupby('Category')['Cholesterol (% Daily Value)'].nunique()).sort_values(by = 'Cholesterol (% Daily Value)' , ascending = False)
chol
```

```python
plt.figure(figsize=(15,10))
plt.pie(chol['Cholesterol (% Daily Value)'] , labels = chol['Cholesterol (% Daily Value)'].index , autopct = '%1.1f%%' , explode = (0.30,0,0,0,0,0,0,0,0));
```

### Category breakfast contribute maximum % of cholesterol (% daily value ) in a diet 

```python
sod = pd.DataFrame(mc.groupby('Item')['Sodium'].sum()).sort_values(by = 'Sodium' , ascending = False).head()
sod
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=sod , y= 'Sodium' , x= sod.index);
plt.xticks (rotation = 90);
plt.title('Items with highest Sodium content' , fontsize  = 20);
```

### Item that contributes maximum to sodium intake is ' Chicken McNuggets (40 piece)	'

```python
satfat = pd.DataFrame(mc.groupby('Item')['Saturated Fat'].sum()).sort_values(by = 'Saturated Fat' , ascending = False).head(10)
satfat
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=satfat , y= 'Saturated Fat' , x= satfat.index);
plt.xticks (rotation = 90);
plt.title('Items with most Saturated Fats ', fontsize = 20);
```
