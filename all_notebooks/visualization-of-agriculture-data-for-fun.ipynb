```python
percentages=['Coarse wool price % Change',
       'Copra price % Change', 
       'Cotton price % Change', 'Fine wool price % Change',
       'Hard log price % Change', 
       'Hard sawnwood price % Change', 'Hide price % change',
       'Plywood price % Change', 
       'Rubber price % Change', 'Softlog price % Change',
       'Soft sawnwood price % Change',
       'Wood pulp price % Change']
```

Correlation Matrix

```python
corrmat = df.corr()
fig = plt.figure(figsize = (12, 9))

sns.heatmap(corrmat, vmax = .8, square = True, annot = True)
plt.show()
```

```python
df=df.replace('-', '0')
```

```python
for i in percentages:
    df[i]=df[i].str.replace('%', '')
    df[i]=df[i].astype('float')
```

```python
df.describe()
```

```python
prices=['Coarse wool Price',
       'Copra Price', 'Cotton Price',
       'Fine wool Price',
       'Hard log Price', 'Hard sawnwood Price',
       'Hide Price', 
       'Plywood Price', 'Rubber Price',
       'Softlog Price', 
       'Soft sawnwood Price', 
       'Wood pulp Price']
```

```python
colors=['#b84949', '#ff6f00', '#ffbb00', '#9dff00', '#329906', '#439c55', '#67c79e', '#00a1db', '#002254', '#5313c2', '#c40fdb', '#e354aa']
```

Univariate Analysis

```python
#univariate analysis
import plotly.express as px
x=0
for i in prices:
    #df = px.data.tips()
    fig = px.histogram(df, x=i, nbins=100, opacity=0.8,
                   color_discrete_sequence=[colors[x]])
    fig.show()
    x+=1
```

Time Series Visualization

```python
import plotly.graph_objs as go
import plotly.offline as pyoff
x=0
for i in prices:
    plot_data = [
        go.Scatter(
            x=df['Month'],
            y=df[i],
            name=i,
            marker = dict(color = colors[x])
            #x_axis="OTI",
            #y_axis="time",
        )
    ]
    plot_layout = go.Layout(
            title=i,
            yaxis_title=i,
            xaxis_title='Month'
        )
    fig = go.Figure(data=plot_data, layout=plot_layout)
    pyoff.iplot(fig)
    x+=1
```

```python
import plotly.graph_objs as go
import plotly.offline as pyoff
x=0
for i in percentages:
    plot_data = [
        go.Scatter(
            x=df['Month'],
            y=df[i],
            name=i,
            marker = dict(color = colors[x])
            #x_axis="OTI",
            #y_axis="time",
        )
    ]
    plot_layout = go.Layout(
            title=i,
            yaxis_title=i,
            xaxis_title='Month'
        )
    fig = go.Figure(data=plot_data, layout=plot_layout)
    pyoff.iplot(fig)
    x+=1
```

Overlapped Graphs for Comparision

```python
import plotly.graph_objs as go
import plotly.offline as pyoff
x=0
plot_data=[]
for i in prices:
    item= go.Scatter(
            x=df['Month'],
            y=df[i],
            name=i,
            marker = dict(color = colors[x])
            #x_axis="OTI",
            #y_axis="time",
        )
    plot_data.append(item)
    x+=1
plot_layout = go.Layout(
        title='Overlapped Prices',
        #yaxis_title=i,
        xaxis_title='Month'
    )
fig = go.Figure(data=plot_data, layout=plot_layout)
pyoff.iplot(fig)
```
