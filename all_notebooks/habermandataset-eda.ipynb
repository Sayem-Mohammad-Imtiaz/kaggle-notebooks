# **Univariate analysis:**

Age survived or not:

```python
import warnings 
warnings.filterwarnings("ignore") 
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="survived_after_5_years", size=4) \
   .map(sns.distplot, "age") \
   .add_legend()
plt.ylabel("Density")
plt.title("Age on survival")
plt.show()
```

**observation:**

There is not much to difference based on age

There is Risk of cancer death above the age of 40



### Year survived or not:

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="survived_after_5_years", size=4) \
   .map(sns.distplot, "year") \
   .add_legend()
plt.ylabel("Density")
plt.title("Time(years) on survival")
plt.show()
```

**observation:**

There is not much to difference based on year

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="survived_after_5_years", size=4) \
   .map(sns.distplot, "lymph_nodes") \
   .add_legend()
plt.ylabel("Density")
plt.title("lymph_nodes on survival")
plt.show()
```

```python
df.lymph_nodes.median()
```

**Observation:**

Here the survival rate is more if the number of lymph_nodes are less

# PDF and CDF

```python
counts, bin_edges = np.histogram(df["age"], bins=10, density=True)
   
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf, label="pdf")
plt.plot(bin_edges[1:], cdf, label = "cdf")
plt.legend()
plt.ylabel("Percentage of person's")
plt.xlabel("age")
plt.title("Plotting CDF and PDF for Age")
plt.show()
```

```python
df.age.min()
```

**Observation:**

According to the PDF, the Risk of having cancer increases above 40 to 55 and then, after 57 gradually decrease
    

```python
counts, bin_edges = np.histogram(df["year"], bins=10, density=True)
   
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf, label="pdf")
plt.plot(bin_edges[1:], cdf, label = "cdf")
plt.xlabel("year")
plt.ylabel("Percentage of person's")
plt.title("Plotting CDF and PDF for Years or time")
plt.legend()
plt.show()
```

```python
df.year.min()
```

```python
counts, bin_edges = np.histogram(df["lymph_nodes"], bins=10, density=True)
   
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf, label="pdf")
plt.plot(bin_edges[1:], cdf, label = "cdf")
plt.legend()
plt.xlabel("lymph_nodes")
plt.ylabel("Percentage of person's")
plt.title("Plotting CDF and PDF for lymph_nodes")
plt.show()
```

### **Observation:**

1. There are more than 80 % of the patients with less than 10 lymph_nodes and most of them survived

2. As the survival rate is higher with less lymph node in the univariate analysis based on lymph_node

# Multivariate Analysis

Using age and lymph_nodes 

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="survived_after_5_years", size=5) \
   .map(plt.scatter, "age", "lymph_nodes") \
   .add_legend();
plt.title("Multivariate analysis of Age and Lymp nodes")
plt.xlabel("Age")
plt.ylabel("Lymp nodes")

plt.show();
```

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="survived_after_5_years", size=5) \
   .map(plt.scatter, "age", "year") \
   .add_legend();
plt.title("Multivariate analysis of Age and Year")
plt.xlabel("Age")
plt.ylabel("Year")
plt.show();
```

# **Conclusion:**

1. Age from 30 - 40 there are more people survived, and lymph nodes doesnot matter they had more "survival chances".
2. As the Age increases from 40 the deaths are more even with less than 10 lymph nodes.
3. Age playing a vital Role for the survivabiltiy even when the lymph nodes are more than 10.
