```python
df = pd.read_csv("/kaggle/input/bitcoin-price-and-m2-money-supply/m2_and_btc.csv")
df.info()
```

```python
# Format date
df["date"] = pd.to_datetime(df["date"], format="%d/%m/%Y")

df = df.set_index("date")

print(df.tail())
```

```python
# Check correlation
df.corr()
```

We can already see a strong correlation of 83% between M2 money and the bitcoin price

```python
import matplotlib
import matplotlib.style
import matplotlib.pyplot as plt

# BTC PRICE VS M2
# creating axes object and defining plot
ax = df.plot(figsize = (14,7), kind='line',
                  y="btc/U$ price", color='orange',
                  linewidth=2)

ax2 = df.plot(kind='line',
                   y='m2(U$ bn)', secondary_y=True,
                   color='green', linewidth=2,
                   ax=ax)

# title of the plot
plt.title("bitcoin price and m2 money supply - correlation : 83%")

# labeling x and y-axis
ax.set_xlabel('Date')
ax.set_ylabel('btc price U$', color="orange")
ax2.set_ylabel('m2 money supply U$ billions', color='g')

ax.yaxis.set_major_formatter(matplotlib.ticker.StrMethodFormatter('{x:,.0f}'))
ax2.yaxis.set_major_formatter(matplotlib.ticker.StrMethodFormatter('{x:,.0f}'))

# show plot
plt.show()
```
