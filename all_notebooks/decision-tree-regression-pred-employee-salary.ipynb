A Decision Tree Regression model for predicting salary of a employee according to their level, more level means higher the position of the employee and hence higher the salary.
Dataset consists of three features, first feature will be ignored because position or level is one of the same thing, hence only level feature will be taken (level is independent and salary is dependent)

```python
#importing libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing dataset
ds = pd.read_csv('../input/position-salary-dataset/Position_Salaries.csv')
ds.head()
```

```python
#since position column and level column are one of the same thing we'll take only level column as our independent variable
x = ds.iloc[:,1:-1].values
y = ds.iloc[:,-1].values
print(x)
print(y)
```

**SPLITTING THE DATASET**
* as only 10 rows are present in the dataset, no splitting will be done so that we can have more data to train our model
* splitting will decrease accuracy


```python
#training the model
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(x,y)
```

```python
#visualizing predcited result
x_new = np.arange(min(x),max(x),0.1).reshape(-1,1)
plt.scatter(x,y,c = 'red')
plt.plot(x_new,dt.predict(x_new))
plt.title('Decision Tree Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
```

```python
#checking accuracy
from sklearn.metrics import r2_score
r2_score(y,dt.predict(x))
```

ACCURACY CAME 100% REASON BEING DECISION TREE IS PRONE TO OVERFITTING AND IT DOESN'T PERFORM WELL ON SMALL DATASETS.
