# **Top 5 countries in each medal**

```python
def getTop5Countries(col,title):
    top5 = df[['Country',col]].sort_values(by=col,ascending=False).head()
    plt.figure(figsize=(7,7))
    plt.title(title)
    sns.barplot(data=top5,x=col,y='Country')
    plt.show()
    return top5
```

# **Top 5 countries in Gold medal**

```python
getTop5Countries('Gold Medal','Top 5 countries in Gold medal')
```

# **Top 5 countries in Silver medal**

```python
getTop5Countries('Silver Medal','Top 5 countries in Silver medal')
```

# **Top 5 countries in Bronze Medal**

```python
getTop5Countries('Bronze Medal','Top 5 countries in Bronze Medal')
```

# **Country and their total medals**

```python
def getMedals(country):
    country_medals = df[['Gold Medal','Silver Medal','Bronze Medal']][df['Country'] == country]
    colors = ['#ffd700','#cccccc','#cd7f32']
    
    plt.figure(figsize=(7,7))
    country_medals.plot(kind='bar',color=colors)
    
    plt.xlabel('Medals')
    plt.title(f'Medals for {country} ({country_medals.sum().sum()})')
    plt.show()
    
    return country_medals
```

```python
getMedals('United States of America')
```

```python
getMedals('ROC')
```

**Similarly You can get any country medals you want**

# **Ranking Countries Overall**

```python
def getRankOverall(col,title):
    ranks = df[['Country',col,'Rank By Total']].sort_values(by='Rank By Total')
    
    plt.figure(figsize=(20,20))
    sns.barplot(data=ranks,y='Country',x=col)
    
    plt.title(title)
    plt.xscale('log')
    plt.show()
```

```python
getRankOverall('Total','Overall Rankings')
```

# **Overall Rankings and Total Gold Medals**

```python
getRankOverall('Gold Medal','Overall Rankings and Total Gold Medals')
```

# **Overall Rankings and Total Silver Medals**

```python
getRankOverall('Silver Medal','Overall Rankings and Total Silver Medals')
```

# **Overall Rankings and Total Bronze Medal**
