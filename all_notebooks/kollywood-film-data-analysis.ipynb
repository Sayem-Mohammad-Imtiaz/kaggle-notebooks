```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
import warnings as ws
ws.filterwarnings("ignore")
```

```python
df = pd.read_csv("/kaggle/input/kollywood-movie-dataset-2011-2017/Kollywood Movie Dataset (2011 - 2017).csv")
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
def clean_name(name):
    return name.lower().strip().replace(" ", "_")
df.rename(columns = clean_name, inplace = True)
```

```python
df.rating.fillna(5.6, inplace = True)
```

```python
df.isna().sum()
```

```python
df.cast = df.cast.str.lower()
```

## LET's Start

```python
# Genere  Wise representation of movies
plt.figure(figsize = (8,30))
sns.set()
plt.title("Diffrent geners Kollywoord movies", fontdict= {"fontsize": 20})
sns.countplot(y = df["genre"], palette="inferno")
plt.xlabel("Generes")
plt.ylabel("Count")
plt.show()
```

# Top Five genres in Kollywood

```python
top_genere = df.groupby('genre')['title'].count().reset_index().sort_values("title", ascending = False)[:10]
```

```python
# Gender Wise representation
plt.figure(figsize = (15,8))
plt.title("Top 10 Movie Genres", fontdict= {"fontsize": 20})
sns.barplot(x="genre", y= "title",data= top_genere, palette="rainbow")
plt.xlabel("Genre of movie")
plt.ylabel("No. of movies")
plt.show()
```

### **** Looks like the Drama is the most loved genre in the Kollywood

# Now we will get the Most Appeared Artist in the industry

```python
superstar = []
for i in df["cast"]:
    temp=i.split()
    if len(temp[0]) > 2:
        superstar.append(temp[0].replace(",", ""))
```

```python
from collections import Counter
count = Counter(superstar)
```

```python
super_star = pd.Series(count).reset_index().rename(columns = {"index": "superstart_artist", 0 : "movies"}).sort_values(by="movies", ascending= False)[:5]
```

```python
# SuperStar Artist
super_star.superstart_artist = super_star.superstart_artist.str.upper()
plt.figure(figsize = (15,8))
plt.title("Top 5 Movie Artist in kollywood", fontdict= {"fontsize": 20})
sns.barplot(x="superstart_artist", y= "movies",data= super_star, palette="pastel")
plt.xlabel("Artist")
plt.ylabel("No. of movies")
plt.show()
```
