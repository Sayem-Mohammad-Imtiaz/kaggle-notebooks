```python

filename = '/kaggle/input/top50spotify2019/top50.csv' #locating the file from the available directory
df = pd.read_csv(filename, encoding = 'ISO-8859-1',index_col=0) #read the CSV file with specific encoding; because without this encoding, the file cannot be read properly
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.countplot(data=df,y='Genre')
#(dance pop) is the most liked genre in top 50 spotify songs. 
```

```python
fig=plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
df.describe().round(decimals=2)
```

```python
df1 = df[['Energy','Loudness..dB..','Beats.Per.Minute','Speechiness.','Valence.']] #create a new dataframe 'df1' which only focus on five parameters
df1.head()
```

```python
dfnorm = (df1-df1.min())/(df1.max()-df1.min())
print(dfnorm.describe())
```

```python
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True, figsize=(10,5))

axs[0].hist(df['Speechiness.'], bins=5) #for non-normalized data
axs[1].hist(dfnorm['Speechiness.'], bins=5) #for normalized data
```

```python
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True, figsize=(10,5))

axs[0].hist(df['Energy'], bins=5) #for non-normalized data
axs[1].hist(dfnorm['Energy'], bins=5) #for normalized data
```

```python
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True, figsize=(10,5))

axs[0].hist(df['Beats.Per.Minute'], bins=5) #for non-normalized data
axs[1].hist(dfnorm['Beats.Per.Minute'], bins=5) #for normalized data
```

```python
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True, figsize=(10,5))

axs[0].hist(df['Loudness..dB..'], bins=5) #for non-normalized data
axs[1].hist(dfnorm['Loudness..dB..'], bins=5) #for normalized data
```

```python
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True, figsize=(10,5))

axs[0].hist(df['Valence.'], bins=5) #for non-normalized data
axs[1].hist(dfnorm['Valence.'], bins=5) #for normalized data
```
