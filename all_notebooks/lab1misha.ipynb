```python
'''
ИМПОРТИРУЕМ НУЖНЫЕ ДЛЯ ЛАБЫ БИБЛИОТЕКИ
'''
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.preprocessing import OneHotEncoder
import warnings
warnings.filterwarnings("ignore")
```

```python
'''
Скачиваем данные и показываем первые 5 строк
'''
data = pd.read_csv('../input/happiness-and-alcohol-consumption/HappinessAlcoholConsumption.csv')
data.head()
```

```python
'''
Вывод графика, который показывает, что данные полные, нет пропусков
'''
sns.heatmap(pd.isnull(data))
```

```python
'''
Так как в задаче регрессии нам не нужны уникальные элементы такие как Country и Region(не совсем уникальный)
То просто удаляем две эти колонки
Далее видим, что в Hemisphere есть north и noth, очевидно - это опечатка и нужно ее пофиксить,
Поэтому заменяем все noth на north (игнорируем предупреждения)
'''
data = data.drop('Country', axis=1)
data = data.drop('Region', axis=1)
print(data.Hemisphere.unique())
data.Hemisphere[data.Hemisphere == 'noth'] = 'north'
print(data.Hemisphere.unique())
data.describe() # выводим статистику по данным
```

```python
'''
Рисуем графики зависимости данных: колонка относительно колонки относительно Hemisphere
'''
sns.pairplot(data, hue='Hemisphere')
```

```python
'''
График зависимости данных(немного отличается от предыдущего и показывает коэффициент корелляции Пирсона
между колонками данных) ''' 
sns.heatmap(data.corr())
```

```python
'''
Преобразовываем буквенные значения в числовые(1 и 0)
'''
columns = ['Hemisphere']

dum = OneHotEncoder(sparse=False)
enc = pd.DataFrame(dum.fit_transform(data[columns]))
```

```python
'''
Разделяем данные на X и Y  и на обучающую и тестовую выборкм
'''
X = pd.concat([enc, data[['HDI', 'GDP_PerCapita', 'Beer_PerCapita', 'Spirit_PerCapita', 'Wine_PerCapita']]], axis=1)
Y = data.HappinessScore
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.5)
```

```python
X.head()
```

```python
lasso = linear_model.Lasso(alpha=0.000001, normalize=True, max_iter=2200).fit(X_train, Y_train)
lasso.score(X_test, Y_test)
```

```python
ridge = linear_model.Ridge(alpha=0.01,normalize=True,solver='lsqr').fit(X_train, Y_train)
ridge.score(X_test, Y_test)
```

```python
elnet = linear_model.ElasticNet(alpha=0.00001,l1_ratio=1, normalize=True, max_iter=1500).fit(X_train, Y_train)
elnet.score(X_test, Y_test)
```

```python
from sklearn.cluster import KMeans
```

```python
data_for_clustering = data.drop('Hemisphere', axis=1)
inert = []
for i in range(1,10):
    km = KMeans(n_clusters=i).fit(data_for_clustering)
    inert.append(km.inertia_)
plt.plot(inert)
plt.show()
#Как видно на графике, лучше всего выбрать к = 2
```

```python
km = KMeans(n_clusters=2).fit(data_for_clustering)
print(km.cluster_centers_)
```

```python
data_for_clustering['Cluster'] = km.labels_ + 1
sns.pairplot(data_for_clustering, hue='Cluster')
```
