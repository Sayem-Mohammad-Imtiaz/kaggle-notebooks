One function that is important to know and how to use is the .mean() function in Pandas. Given in the name, this function allows you to calculate the mean. 

Given a set of rows and columns let's take a look at how the mean function in pandas would look down below: 

```python
import pandas as pd

df =  pd.DataFrame({"A":[6, 9, 12, 7],
                   "B":[3, 4, 11, 1], 
                   "C":[11, 6, 15, 15],
                   "D":[15, 5, 7, 2]})
print("Example Data Below:")
print(df)
df.mean()
```

As seen above, the mean is calculated for every column, which is done by averaging all the row values for that specific column. 
Example: Mean of Column A
Column A: 
 Row 1: 6
 Row 2: 9
 Row 3: 12
 Row 4: 7
 6 + 9 + 12 + 7 = 34
 34 / 4 = 8.50
            

What if we wanted to collect the mean of all the columns given above? 

```python
import pandas as pd

df =  pd.DataFrame({"A":[6, 9, 12, 7],
                   "B":[3, 4, 11, 1], 
                   "C":[11, 6, 15, 15],
                   "D":[15, 5, 7, 2]})
print("Example Data Below:")
print(df)
column_means = df.mean()

total_mean = column_means.mean()
print(column_means)
print("Final Mean:" , total_mean)

```

See what we did above? We took the mean of the column's means by assigning a variable to the given column means. 

There is a specific parameter that is common for a lot of other function you are going to learn in this class called the axis parameter. As seen above, the default is always axis = 0. Axis = 0 is where the mean is calculate for every column averaging all of the row values. 

What if we wanted to take the mean of an individual row and average all the column values? 

```python
import pandas as pd

df =  pd.DataFrame({"A":[6, 9, 12, 7],
                   "B":[3, 4, 11, 1], 
                   "C":[11, 6, 15, 15],
                   "D":[15, 5, 7, 2]})
print("Example Data Below:")
print(df)
df.mean(axis = 1)
```

When we changed the axis parameter to equal 1, we get the average for every row using the column values.

As we have learned before, you can clean data and removed values which you do not want to have in your dataset. What if we were to have data that is not cleaned? 

```python
import pandas as pd
   
df = pd.DataFrame({"A":[9, 4, 5, None, 6],
                   "B":[5, 6, 13, 3, None],
                   "C":[20, 16, 11, 3, 7],
                   "D":[15, 5, None, 2, 6]})
  
# finding the mean of every row while skipping the null values
print(df)
df.mean(axis = 1, skipna = True)
```

As shown in the example above, we use the second parameter where skipna = True. This indicates that we want to skip the Null values when calculating the median. 

Notice that we make skipna = True, but what if we wanted to take into account the Null values? We would simply just changed skipna = False. This will be important while using functions later on.

Now, let's move on to seeing the .min() and .max() functions which, given in their names represent the minimum and maximum values of a dataset. 

There are many different scenarios in which you can find the minimum and maximum, and we are going to show you the different scenarios, giving you an idea on how to play around with the parameters of these pandas functions. 

```python
import pandas as pd
  
data = [
          (20, 16, 23),
          (30, None, 11),
          (40, 34, 11),
          (50, 35, None),
          (60, 40, 13)
        ]
  
# creating a DataFrame object

df = pd.DataFrame(data)
  

# minimum value from each COLUMN
minimum_value = df.min()
print(df)  
print(minimum_value)
```

The .min() function is very similar to the functions we have seen previously. 

Let's see what other scenarios we can see the .min() function be used! 

```python
import pandas as pd
  
data = [
          (20, 16, 23),
          (30, None, 11),
          (40, 34, 11),
          (50, 35, None),
          (60, 40, 13)
        ]
  
# creating a DataFrame object

df = pd.DataFrame(data)
  

# minimum value from each ROW
minimum_value = df.min(axis = 1)
print(df)  
print(minimum_value)
```

```python
import pandas as pd
  
data = [
          (20, 16, 23),
          (30, None, 11),
          (40, 34, 11),
          (50, 35, None),
          (60, 40, 13)
        ]
  
# creating a DataFrame object

df = pd.DataFrame(data)
  

# minimum value from each ROW
minimum_value = df.min(axis = 1)
print(df)  
print(minimum_value)
```

There is a slight difference here, we wanted the minimum value from the row instead of the column. Therefore, we used axis = 1, instead of leaving the parameter for the .min() function blank. REMEMBER: When no parameter is given for the axis, it defaults to axis = 0!

So for some reason, you are asked to find the minimum without skipping the Null values and to find the minimum value of each row instead of column, what do you do?

```python
import pandas as pd
  
data = [
          (20, 16, 23),
          (30, None, 11),
          (40, 34, 11),
          (50, 35, None),
          (60, 40, 13)
        ]
  
# creating a DataFrame object

df = pd.DataFrame(data)
  

# minimum value from each ROW while not skipping the Null Values!
minimum_value = df.min(axis = 1, skipna = False)
print(df)  
print(minimum_value)
```

The great thing about the .min() function is that, the .max() function acts exactly the same with the same basic parameters! 

```python
import pandas as pd
  
data = [
          (20, 16, 23),
          (30, None, 11),
          (40, 34, 11),
          (50, 35, None),
          (60, 40, 13)
        ]
  
# creating a DataFrame object

df = pd.DataFrame(data)
  

# maximum value from each ROW while skipping the Null Values!
max_value = df.min(axis = 1, skipna = True)
print(df)  
print(max_value)
```

Now we are going to go over the .quantile() function in Pandas! This function returns the quantile at any given axis. 

What if we want to find the 50% quantile?

```python
import pandas as pd

data = {"A":[9, 4, 5, None, 6],
                   "B":[5, 6, 13, 3, None],
                   "C":[20, 16, 11, 3, 7],
                   "D":[15, 5, None, 2, 6]}
  
df = pd.DataFrame(data)
  
quantile_value = df.quantile()

print(df)
print(quantile_value)

  

```

As seen above, the 50% quantile is the default value, so no parameter needs to be given if you wanted to find the 50% quantile!

Suppose we wanted to find the 10%, 25%, 50%, and 75% quantiles along the index axis. 

```python
data = {"A":[9, 4, 5, None, 6],
                   "B":[5, 6, 13, 3, None],
                   "C":[20, 16, 11, 3, 7],
                   "D":[15, 5, None, 2, 6]}
  
df = pd.DataFrame(data)

quantile_values = df.quantile([0.1,0.25,0.50,0.75], axis = 0)
# NOTE: you do not need to include the axis = 0 parameter, however it makes it clear for the readers or clearer for yourself
print(quantile_values)
```

The .var() calculates variance of a dataset, and has the same basic parameters as most of the functions mentioned above (skipna, and axis). However, because variance is a scalar concept (one-dimensional) the only axis available is the default axis. 

For this reason, when we create the data, we use pd.Series instead of pd.DataFrame because Series indicates that there can only be a single list with index, whereas a DataFrame can contain multiple series.

```python
import pandas as pd
  
data = pd.Series([19.5, 16.8, 22.78, 20.124, 18.1002])
# Notice the use of .Series instead of .DataFrame
print(data)
print(data.var())
```

```python
# TRY IT!
# Question: Find the variance of the data given while excluding the Null values. 

import pandas as pd

data = pd.Series([100, 214, 325, 88, None, 325, None, 68])
  
df = pd.Series(data)

#Teacher version include this
variance = df.var(skipna = True)

print(variance)
```
