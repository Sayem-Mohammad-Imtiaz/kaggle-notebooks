**Gráfique las siguientes funciones con matplotlib:**
* logaritmo natural (de 1 a 10) (1 pts)
* exponencial (de -5 a 5) (1pts)
* tangente hiperbólico (de -10 a 10) (1 pts)
* Todas las funciones deben estár en la misma línea

```python
import matplotlib.pyplot as plt
import numpy as np
import math

log_numbers = np.arange(1,11)
exp_numbers = np.arange(-5,6)
tanh_numbers = np.arange(-10,11)
plt.figure(figsize=(9, 3))
plt.subplot(131)
plt.plot(log_numbers, [np.log(number) for number in log_numbers])
plt.subplot(132)
plt.plot(exp_numbers, [math.exp(number) for number in exp_numbers])
plt.subplot(133)
plt.plot(tanh_numbers, [math.tanh(number) for number in tanh_numbers])
plt.show()
```

**El siguiente dataset contiene una serie de cualidades de los vinos con las cuales se les otorga un puntaje (quality) de 0 a 10. El dataset original está disponible en: Red Wine Quality.** 
* Utilice un gráfico de dispersión (Scatter) para graficar en el eje x el pH
 en el eje y el alcohol
* el color de los puntos debe ser la calidad (quality) 
* y debe incluir los títulos de los ejes así como el título del gráfico. (3 pts)

```python
import pandas as pd
import matplotlib.pyplot as plt
wines = pd.read_csv('../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
wines.dropna()
#wines.plot.scatter(x='pH', y='alcohol', c=wines["quality"])

plt.scatter( # usar .scatter permite configurar características propias de cada gráfico
    wines["pH"], 
    wines["alcohol"], 
    c=wines["quality"],
    alpha=0.3, # transparencia del punto
    cmap='viridis' # mapa del color
)
plt.title('Calidad del vino según su pH y nivel de Alcohol')
plt.colorbar(); # muestra la escala de color
```

**Genere un gráfico de líneas utilizando plotly. Utilice el dataset àvocados que se encuentra a continuación.** 
* Para el eje x utilice la fecha (Date), para el eje y el precio promedio (AveragePrice) y para el color de las líneas utilice la región (Region). 
* ¡Solamente debe graficar los precios para Los Angeles y San Francisco! (3 pts)

```python
import pandas as pd
import plotly.express as px
avocados = pd.read_csv('../input/avocado-prices/avocado.csv')
avocados["Date"] = pd.to_datetime(avocados['Date'], format='%Y-%m-%d')

avocados = avocados[avocados.region.isin(['LosAngeles', 'SanFrancisco'])]
fig = px.line(avocados, x="Date", y="AveragePrice", color='region', title='Prices for LA and San Francisco!')
fig.show()
```
