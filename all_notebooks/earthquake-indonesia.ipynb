```python
EQ_2017 = pd.read_csv('/kaggle/input/earthquake-around-indonesian-territory/GempaTh2017.csv')  
EQ_2018 = pd.read_csv('/kaggle/input/earthquake-around-indonesian-territory/GempaTh2018.csv')  
EQ_2019 = pd.read_csv('/kaggle/input/earthquake-around-indonesian-territory/GempaTh2019.csv')  
EQ_2020 = pd.read_csv('/kaggle/input/earthquake-around-indonesian-territory/GempaTh2020.csv')  

# Indonesian Territory
min_Lon = 93
max_Lon = 141.25
min_Lat = -15
max_Lat = 9

# Retrieving data based on Indonesian territory
EQ2017_df = EQ_2017.loc[(EQ_2017['Lon'] >= min_Lon) &
                        (EQ_2017['Lon'] <= max_Lon) &
                        (EQ_2017['Lat'] >= min_Lat) &
                        (EQ_2017['Lat'] <= max_Lat)]

EQ2018_df = EQ_2018.loc[(EQ_2018['Lon'] >= min_Lon) &
                        (EQ_2018['Lon'] <= max_Lon) &
                        (EQ_2018['Lat'] >= min_Lat) &
                        (EQ_2018['Lat'] <= max_Lat)]

EQ2019_df = EQ_2019.loc[(EQ_2019['Lon'] >= min_Lon) &
                        (EQ_2019['Lon'] <= max_Lon) &
                        (EQ_2019['Lat'] >= min_Lat) &
                        (EQ_2019['Lat'] <= max_Lat)]

EQ2020_df = EQ_2020.loc[(EQ_2019['Lon'] >= min_Lon) &
                        (EQ_2019['Lon'] <= max_Lon) &
                        (EQ_2019['Lat'] >= min_Lat) &
                        (EQ_2019['Lat'] <= max_Lat)]
```

```python
# create a dataframe based on the total number of earthquakes per year
Y_EQ = ["2017", "2018", "2019", "2020" ]
No_EQ = np.array([len(EQ2017_df), len(EQ2018_df),len(EQ2019_df),len(EQ2020_df)], dtype=int)

data = {"Year": Y_EQ,
       "Num": No_EQ}

# The number of earthquakes every year
No_EQ = pd.DataFrame(data = data)
No_EQ
```

```python
# make a graph based on the number of earthquakes
plt.subplots(figsize=(16, 8))

plt.plot(No_EQ['Year'],No_EQ['Num'],'--ok')
rect1 = plt.bar(No_EQ['Year'],No_EQ['Num'],0.4)
plt.grid(True)
plt.ylabel("num")
plt.xlabel("year")
plt.title("Number of earthquakes around the Indonesian Territory")

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        plt.annotate('{}'.format(height),
                     xy=(rect.get_x() + rect.get_width() / 2, height),
                     xytext=(0, 3),  # 3 points vertical offset
                     textcoords="offset points",
                     ha='center', va='bottom')
autolabel(rect1)      

plt.show()
```

```python
# create a dataframe based on the earthquake magnitude range
index = ["3 - <4", "4 - <5", "5 - <6", " =>6"]

No_Mag_2017 = np.zeros(len(index), dtype=int)
No_Mag_2018 = np.zeros(len(index), dtype=int)
No_Mag_2019 = np.zeros(len(index), dtype=int)
No_Mag_2020 = np.zeros(len(index), dtype=int)

No_Mag_2017[0] = EQ2017_df['Mag'][(EQ2017_df['Mag'] >= 3) & (EQ2017_df['Mag'] < 4) ].count()
No_Mag_2017[1] = EQ2017_df['Mag'][(EQ2017_df['Mag'] >= 4) & (EQ2017_df['Mag'] < 5) ].count()
No_Mag_2017[2] = EQ2017_df['Mag'][(EQ2017_df['Mag'] >= 5) & (EQ2017_df['Mag'] < 6) ].count()
No_Mag_2017[3] = EQ2017_df['Mag'][(EQ2017_df['Mag'] >= 6) & (EQ2017_df['Mag'] < 10) ].count()

No_Mag_2018[0] = EQ2018_df['Mag'][(EQ2018_df['Mag'] >= 3) & (EQ2018_df['Mag'] < 4) ].count()
No_Mag_2018[1] = EQ2018_df['Mag'][(EQ2018_df['Mag'] >= 4) & (EQ2018_df['Mag'] < 5) ].count()
No_Mag_2018[2] = EQ2018_df['Mag'][(EQ2018_df['Mag'] >= 5) & (EQ2018_df['Mag'] < 6) ].count()
No_Mag_2018[3] = EQ2018_df['Mag'][(EQ2018_df['Mag'] >= 6) & (EQ2018_df['Mag'] < 10) ].count()

No_Mag_2019[0] = EQ2019_df['Mag'][(EQ2019_df['Mag'] >= 3) & (EQ2019_df['Mag'] < 4) ].count()
No_Mag_2019[1] = EQ2019_df['Mag'][(EQ2019_df['Mag'] >= 4) & (EQ2019_df['Mag'] < 5) ].count()
No_Mag_2019[2] = EQ2019_df['Mag'][(EQ2019_df['Mag'] >= 5) & (EQ2019_df['Mag'] < 6) ].count()
No_Mag_2019[3] = EQ2019_df['Mag'][(EQ2019_df['Mag'] >= 6) & (EQ2019_df['Mag'] < 10) ].count()

No_Mag_2020[0] = EQ2020_df['Mag'][(EQ2020_df['Mag'] >= 3) & (EQ2020_df['Mag'] < 4) ].count()
No_Mag_2020[1] = EQ2020_df['Mag'][(EQ2020_df['Mag'] >= 4) & (EQ2020_df['Mag'] < 5) ].count()
No_Mag_2020[2] = EQ2020_df['Mag'][(EQ2020_df['Mag'] >= 5) & (EQ2020_df['Mag'] < 6) ].count()
No_Mag_2020[3] = EQ2020_df['Mag'][(EQ2020_df['Mag'] >= 6) & (EQ2020_df['Mag'] < 10) ].count()

No_Mag_df = pd.DataFrame({"2017": No_Mag_2017,
                          "2018": No_Mag_2018,
                          "2019": No_Mag_2019,
                          "2020": No_Mag_2020}, index= index)

No_Mag_df
```
