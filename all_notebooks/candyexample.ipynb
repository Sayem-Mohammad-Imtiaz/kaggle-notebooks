Let's now do some useful thing!!  Try to find the candy with the highest sugar level - 95%.



```python
candy_data = pd.read_csv('/kaggle/input/fivethirtyeight-candy-power-ranking-dataset/candy-data.csv')

#find all candies with sugar percent value of 95% or more.
condition_candy_data = candy_data[(candy_data.sugarpercent >= 0.95)]
#Now sort it in descending order
sorted_candy_data = condition_candy_data.sort_values(by=['sugarpercent'], ascending=False)
#Show up to top 10 data points
sorted_candy_data.head(10)
```
