```python
import pandas as pd
import numpy as np
import keras 
from keras import layers
from keras.models import Model
from keras.optimizers import SGD
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv('../input/housesalesprediction/kc_house_data.csv')
```

```python
#Selection des donnees
df=df[['sqft_lot','floors','price','bathrooms','bedrooms','sqft_living']]
df_x=df[['sqft_lot','floors','bathrooms','bedrooms','sqft_living']]
df_y=df[['price']]
```

```python
#pretraitement des donnees
scaler=MinMaxScaler()

scaler.fit(df_x)
new_df_x=scaler.transform(df_x)

scaler.fit(df_y)
new_df_y=scaler.transform(df_y)
```

```python
#Separation des donnees
#donnees d'apprentissage
train_x=new_df_x[0:17000]
train_y=new_df_y[0:17000]

#donnees de test
test_x=new_df_x[17000:]
test_y=new_df_y[17000:]

print(train_x.shape,train_y.shape)
print(test_x.shape,test_y.shape)
```

```python
#creation de reseau de neurones
inpt=layers.Input(shape=(5,))

x=layers.Dense(100)(inpt)
actv1=layers.Activation('relu')(x)

y=layers.Dense(100)(actv1)
actv2=layers.Activation('relu')(y)

z=layers.Dense(100)(actv2)
actv=layers.Activation('relu')(z)

output=layers.Dense(1)(actv)
model=Model(inpt,output)
```

```python
#summary of model
model.summary()
```

```python

opt=SGD()

#compilation du model
model.compile(optimizer=opt,loss='mse')
```

```python
#lancer l'apprentissage
model.fit(train_x,train_y,epochs=50,batch_size=64)
```

```python

#calcule l'erreur de prediction
predictions=model.predict(test_x)
np.mean((predictions - test_y)**2)
```
