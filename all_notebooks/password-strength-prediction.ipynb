```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('../input/traindata/data.csv',',',error_bad_lines=False)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
y = df['strength']
x = df['password']
```

```python
x.head()
```

```python
y.head()
```

```python
sns.countplot(x='strength',data=df)
```

```python
def wtc(inputs):
    characters=[]
    for i in inputs:
        characters.append(i)
    return characters
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
v=TfidfVectorizer(tokenizer=wtc)
x=v.fit_transform(x)
```

```python
fn=v.get_feature_names()
first_document_vector=x[0]
```

```python
df1 = pd.DataFrame(first_document_vector.T.todense(),index=fn,columns=["tfidf"])
df1.sort_values(by=["tfidf"],ascending=False)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
lc=LogisticRegression(penalty='l2',multi_class='ovr',solver='lbfgs',max_iter=400)
lc.fit(x_train,y_train)
```

```python
print(lc.score(x_test,y_test))
```

```python
import tkinter as tk
from tkinter import *
```

```python
import pickle
model = pickle.dumps(lc)
```

```python
from tkinter import messagebox
import os
if os.environ.get('DISPLAY','') == '':
    print('no display found. Using :0.0')
    os.environ.__setitem__('DISPLAY', ':0.0')

window=Tk()
window.geometry("300x100")
window.title("Password Strength")

pswd=StringVar()

DF=pd.DataFrame()
def action():
    pwd = pswd.get()
    x_pred=np.array([pwd])
    xx_pred=v.transform(x_pred)
    y_pred=lc.predict(xx_pred)
    result=y_pred
    if(result==0 or result==1):
        messagebox.showinfo('Password Strength','Your password is WEAK')
    elif(result==2):
        messagebox.showinfo('Password Strength','Your password is STRONG')

label=Label(window,text="Enter your password here",font=("times new roman",10,"bold"))
label.pack()

e1=Entry(window,textvar=pswd)
e1.place(x=90,y=30)


button1=Button(window,text="Cancel",command=window.destroy)
button1.place(x=110,y=60)

button2=Button(window,text="OK",command=action)
button2.place(x=160,y=60)

window.mainloop()
```
