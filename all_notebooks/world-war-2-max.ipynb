### Part 2 Data Training,Testing and Predictions.

Now Let's Select Features on which the Maximium Temperature Depends.

```python
worldWar2Data.columns
```

As we are going to check the Relationship between Max and Min Temperature so let's select Min Temperature as Feature.

```python
X_Features = worldWar2Data[['MinTemp']]
```

```python
X_Features
```

Now as we are going to predict the Max Temperature so let's select Max Temperature as a thing to Predict.

```python
Y_Predict = worldWar2Data['MaxTemp']
```

```python
Y_Predict
```

Now Let's import the Train and Test Split form the scikit learn in order to get training and testing sets.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_Features, Y_Predict, test_size=0.40, random_state=101)
```

Now Let's Import the Linear Regression Model in order to train and test it.

```python
from sklearn.linear_model import LinearRegression
```

Initiating the Instance of Linear Regression Model.

```python
linearRegressionInstance = LinearRegression()
```

Now Fitting the Linear Regression Model.

```python
linearRegressionInstance.fit(X_train,y_train)
```

Now let's check the intercept.

```python
linearRegressionInstance.intercept_
```

Now Let's check the Co-Efficients.

```python
linearRegressionInstance.coef_
```

Let's Describe the Coefficients in a more describeable way.

```python
coefficientDescription = pandasInstance.DataFrame(linearRegressionInstance.coef_,index=X_train.columns,columns=['Co-Efficient Values'])
```

```python
coefficientDescription
```

##### Now let's Predict the Max Temp.

```python
maxTempPredictions = linearRegressionInstance.predict(X_test)
```

```python
maxTempPredictions
```

##### Now let's Compare the Predicted and Actual Values on a Scatter Graph.

```python
matplotlibInstance.scatter(y_test,maxTempPredictions)
```

##### as we can see that the Graph has very less noise.

##### Now let's Check the Distributaion of the Residuals.

```python
matplotlibInstance.figure(figsize=(15,10))
matplotlibInstance.tight_layout()
seabornInstance.distplot(y_test - maxTempPredictions)
```

##### We can see that the Distribuation is nearly normal which depects that the Model that we have selected is a good choice.

Let's do some work for the Error.

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,maxTempPredictions)
```

```python
metrics.mean_squared_error(y_test,maxTempPredictions)
```

```python
numpyInstance.sqrt(metrics.mean_squared_error(y_test,maxTempPredictions))
```
