*****Which Country has the highest Suicide rate?*****

```python
year_sui = df[['country','suicides_no']].groupby('country')['suicides_no'].sum().sort_values(ascending=False)
year_sui.plot.bar(figsize=(50,10))
```

```python
plt.figure(figsize=(10,5))
year_sui.head(10).plot.bar()
plt.title('Top 10 Countries With Highest Suicide Rate', fontsize=15)
plt.show()
```

```python
df['year'].nunique()
```

The data has been collected over the span of **38** years

***How Suicide rate has changed over the years***

```python
years = df.groupby('year')['suicides_no'].sum().reset_index(name='count')
plt.figure(figsize=(25,5))
sns.pointplot(data=years, x='year', y='count')
plt.title('Suicide rate Over the Years', fontsize=15)
```

There is decline in suicide rate between the years 1983 and 1984 

***Suicide Rate In Male & Female***

```python
colors=['lightpink','lightblue']
fig = px.bar(x=df['sex'].value_counts().index, y=df['sex'].value_counts(), title='Gender Distribution',height=500, width=500,
             text=(df['sex'].value_counts()/len(df['sex'])*100))
fig.update_traces( textposition='outside', texttemplate='%{text:.2s}%', marker_color = colors)
fig.show()
```

```python
colors = ['lightpink','lightblue']

sex_sui = df[['sex','suicides_no']].groupby('sex')['suicides_no'].sum().sort_index()
print(sex_sui)
fig = px.pie(values = sex_sui, labels =sex_sui)

fig['layout'].update(title='Gender VS Suicide',height=400,)
fig.update_traces(hoverinfo="label+percent",textfont_size=15, marker=dict(colors=colors, line=dict(color='white', width=2)))

fig.show()
```

Suicide rate is higher among Males.

***How suicide rate has changed over the years among males and females.***

```python

year_sex_sui = df[['year','sex','suicides_no']].groupby(['year','sex'])['suicides_no'].sum().reset_index()

fig =px.line(x=year_sex_sui['year'],y=year_sex_sui['suicides_no'], color= year_sex_sui['sex'], height=400, width=900, title='Suicide rate In Males and Females over the Years')
fig.show()
```

***Suicide rate among different age groups.***

```python
colors=['mediumturquoise','lightgreen','seagreen',"rgb(114, 78, 145)",'palegreen','olive']

age_sui = df[['age','suicides_no']].groupby('age')['suicides_no'].sum().sort_values( ascending=False)

fig = px.bar(x=age_sui.index, y=age_sui, height=500, width=600,text=(age_sui/len(age_sui)*100), )
fig.update_traces(textposition='outside', texttemplate='%{text:.3s}%', marker_color=colors)
fig['layout'].update(title='Suicide rate  Distribution Among Different Age Groups')
fig.show()
```

Highest suicide rate is between **35-54** ages.

***How suicide rate has changed among different age groups over the years.***

```python
age_year_sui = df[['year','age','suicides_no']].groupby(['year','age'])['suicides_no'].sum().reset_index()

fig = px.line(x=age_year_sui['year'], y=age_year_sui['suicides_no'], color=age_year_sui['age'], title='Suicide rate Among Different Age Groups Over the Years')
fig.show()
```
