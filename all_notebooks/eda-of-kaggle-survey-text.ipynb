```python
WorkAlgorithmsSelect_text = ' '.join(kaggle_survey['WorkAlgorithmsSelect'].dropna().tolist())
# kaggle_survey['WorkAlgorithmsSelect'].dropna()
```

```python
WorkToolsSelect_text = ' '.join(kaggle_survey['WorkToolsSelect'].dropna().tolist())
```

```python
# Create and generate a word cloud image:
plt.subplots(figsize=(12,6))
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
wordcloud = WordCloud().generate(WorkToolsSelect_text)

# Display the generated image:

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
