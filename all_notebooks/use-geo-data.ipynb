```python
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

ds=pd.read_csv("../input/forbes-billionaires-2021-30/forbes_billionaires_geo.csv")
```

```python
from shapely import wkt

ds['geometry'] = ds['geometry'].apply(wkt.loads)
```

```python
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ds = gpd.GeoDataFrame(ds).set_geometry('geometry')
```

```python
base = world.plot(color='white', edgecolor='black', figsize=(16, 9))

ds.plot(ax=base, marker='o', color='magenta', markersize=1);
```

```python
resid = ds.groupby("Residence").count().Name[(ds.groupby("Residence").count().Name>1).values].sort_values()
```

```python
res_plot = ds[ds.Residence.isin(resid.index)]
res_plot = gpd.GeoDataFrame(res_plot.groupby("Residence").count().Name)
```

```python
cc=[]
for i in res_plot.index:
    cc.append(ds[ds.Residence==i].reset_index().geometry[0])
    
res_plot["geometry"]=cc
res_plot.set_geometry('geometry')
```

```python
base = world.plot(color='white', edgecolor='gray', figsize=(35, 15))

ax = res_plot.plot(ax=base, marker='o', cmap='Pastel1', markersize=res_plot.Name*20);
```
