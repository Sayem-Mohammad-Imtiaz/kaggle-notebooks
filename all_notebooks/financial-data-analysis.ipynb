```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import sklearn 
```

```python
df=pd.read_csv('/kaggle/input/finance-data/Finance_data.csv')
df.head()
```

```python
df.keys()  # get all keys of dataset
```

```python
plt.figure(figsize=(12, 7))
sns.countplot(df['gender'],linewidth=3,edgecolor='black')  #number of male and female investers
plt.show()
```

```python
plt.figure(figsize=(12, 7))  #custom size of graph
sns.countplot(x=df['gender'],hue=df['What are your savings objectives?'],linewidth=3,edgecolor='black')
plt.title("Reasons for savings by men and women respectively")
plt.show()
```

### Conclusion: The most obvious reason for both male and female for savings is reitirement plan

```python
plt.figure(figsize=(12, 7))
sns.countplot(df['age'],hue=df['gender'], edgecolor='black',linewidth=2)  #double bar graph for age of investments with hue as gender.
plt.title("Count of investers by age")
plt.figure(figsize=(12, 7))
plt.show()
```

### Conclusion: People around 25 to 30 who invest are more likely to be males

```python
plt.figure(figsize=(12, 7))
sns.countplot(x=df['gender'], hue=df['Investment_Avenues'], edgecolor='black', linewidth=4)
plt.title("Counting whether particular gender took investment avanue or not")
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.jointplot(x=df['age'],y=df['Mutual_Funds'],color='green',hue=df['gender']) #analyzing mutual funds dependancy on age with hue as gender
plt.show()
```

### Conclusion: Age is not correlated with mutual funds investment

```python
plt.figure(figsize=(12, 7))
sns.jointplot(x=df['age'],y=df['Mutual_Funds'],kind='kde',color='green',hue=df['gender']) #analyzing mutual funds dependancy on age using kernel density estimation
plt.show()
```
