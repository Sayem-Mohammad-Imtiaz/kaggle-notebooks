# Neural Network Implementation

## Data Prep

```python
df.shape
```

```python
data = df.iloc[:, :-1]
targets = df.iloc[:, -1]
```

```python
data.head()
```

```python
targets.head()
```

## Splitting Data

```python
train_data, test_data, train_targets, test_targets = train_test_split(data, targets, test_size=0.2)
```

# Standardization

```python
mean = train_data.mean()
std = train_data.std()

train_data -= mean
train_data /= std

test_data -= mean
test_data /= std
```

```python
train_data.head()
```

```python
test_data.head()
```

# Encoding Prediction Target

```python
train_targets = pd.get_dummies(train_targets)
```

```python
test_targets = pd.get_dummies(test_targets)
```

```python
train_targets.head()
```

```python
test_targets.head()
```

# Building Model

```python
model = models.Sequential()

model.add(layers.Dense(10, input_shape=(train_data.shape[1],), activation='relu'))

model.add(layers.Dense(3, activation='softmax'))

model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(train_data, train_targets, epochs=100, batch_size=16, validation_split=0.15)
```

```python
plt.figure(figsize=(10, 10))
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
epochs = range(1, 101)
plt.plot(epochs, acc, label='Accuracy', linewidth=5);
plt.plot(epochs, val_acc, label='Validation Accuracy', linewidth=8);
plt.legend();
plt.xlabel('Epochs');
plt.ylabel('Accuracy');
plt.title('Train and Validation Accuracy');
```

-> model starts overfitting after 40 epochs

# Final Model

```python
model = models.Sequential()

model.add(layers.Dense(10, input_shape=(train_data.shape[1],), activation='relu'))

model.add(layers.Dense(3, activation='softmax'))

model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(train_data, train_targets, epochs=40, batch_size=16, validation_split=0.15)
```

# Evaluation

```python
eval_ = model.evaluate(test_data, test_targets)
```

```python
print(f'Evaluation Accuracy is {((eval_[1])*100):.2f}%')
```
