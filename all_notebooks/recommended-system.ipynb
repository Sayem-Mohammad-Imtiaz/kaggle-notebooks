Now we are going to build Simple Recommended System

```python
def get_index_by_name(name):
    return df[df.name==name].index.values[0]

def get_name_by_index(index):
    return df[df.index==index]['name'].values[0]
```

```python
features = ['diet','ingredients','course','flavor_profile','state']

for feature in features:
    df[feature].fillna(' ')
```

```python
def combine_row(row):
    return row['diet']+ " " + " "+row['ingredients']+ " " +row['course']+ " " +row['flavor_profile']+ " " +row['state']

dff= df.apply(combine_row,axis=1)
```

```python
dff
```

```python
cv = CountVectorizer ()
co_vec = cv.fit_transform(dff)
```

```python
print(enumerate (co_vec))
```

```python
co_sim= cosine_similarity(co_vec)
```

```python
liked_food= 'Red Rice'
i = get_index_by_name(liked_food)
```

```python
sim = list(enumerate (co_sim[i]))
```

```python
simm = sorted(sim,key=lambda x:x[1], reverse=True)
```

```python
i = 0
for movie in simm:
    print (get_name_by_index(movie[0]))
    i = i+1
    if i>20:
        break
```
