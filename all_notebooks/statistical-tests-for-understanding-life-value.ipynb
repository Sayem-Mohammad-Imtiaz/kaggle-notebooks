```python
df_nn['Palestinians Injuries']=df_nn['Palestinians Injuries'].replace("(incl. Jun)", np.nan)
df_nn['Israelis Injuries']=df_nn['Israelis Injuries'].replace("(incl. Jun)", np.nan)
df_nn['Palestinians Injuries']=df_nn['Palestinians Injuries'].replace("(incl. Aug)", np.nan)
df_nn['Israelis Injuries']=df_nn['Israelis Injuries'].replace("(incl. Aug)", np.nan)
```

```python
object_data = ["Palestinians Injuries", "Israelis Injuries", "Palestinians Killed" ,"Israelis Killed"]
for i in object_data:
    df_nn[i] = df_nn[i].astype(float)
    
```

```python
df_nn = df_nn.dropna()
df_nn.info()
```

Based on 130 samples
Max Israelis injured is 2347 and that of Palestinians are 13735
Max Israelis killed is 13 and that of Palestinians are 614.

**Statistical tests**

```python
#understading correlation
import scipy
from scipy.stats import pearsonr
data1 = df_nn["Palestinians Injuries"]
data2 = df_nn["Israelis Injuries"]
data3 = df_nn["Palestinians Killed"]
data4 = df_nn["Israelis Killed"]

stat, p = pearsonr(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably independent')
else:
	print('Probably dependent')
```

High correlation between injuries of Palestinians and Israelis.
When injuries in Palestenians increases, injuries in Israelis increases too.

```python
stat2, p2 = pearsonr(data3, data4)
print('stat=%.3f, p=%.3f' % (stat2, p2))
if p2 > 0.05:
	print('Probably independent')
else:
	print('Probably dependent')
```

Low correlation between Palestinians killed and Israelis killed.

```python
from scipy.stats import ttest_ind
data1 = df_nn["Palestinians Injuries"]
data2 = df_nn["Israelis Injuries"]

stat3, p3 = ttest_ind(data1, data2)
print('stat=%.3f, p=%.3f' % (stat3, p3))
if p3 > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
```

A t score of 3.35 means that the groups are 3.35 times as different from each other as they are within each other. The bigger the t-value, the more likely it is that the results are repeatable.

Our assumption was that the means of the Palestinians injured and mean of Israelis injured are same.
But P-value is less than 0.01, we can say that with 99% confidence that the mean of the injuries between 2 groups are different. We reject our initial assumption or null hypothesis. Less than 1% probability that our initial assumption is correct.
More Palestinians are injured than Israelis.
Life value are getting much lesser for Palestinians. 

```python
#ANOVA
#Tests whether the means of two or more independent samples are significantly different.
from scipy.stats import f_oneway
stat, p = f_oneway(data1, data2, data3, data4)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
```

Means are significantly different among the data as p<0.01

```python
#wilcoxon test
from scipy.stats import wilcoxon
stat, p = wilcoxon(data3, data4)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
```

 we would reject the null hypothesis at a confidence level of 1%, concluding that there is a difference in killed numbers between Palestinians and Israelis. Palestinians are much greater.
