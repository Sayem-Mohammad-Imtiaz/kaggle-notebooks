```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("../input/finance-india/Aggregate_Expenditure.csv",encoding ='latin1')
df
```

```python
rows,columns = df.shape
print("Rows: ",rows)
print("Columns: ",columns)
```

```python
for i in range(columns):
    print(str(i)+". :", df.iloc[:,i].dtype)
```

```python
df.isna().sum()
```

```python
df.replace({"":0},inplace=True)
```

```python
df
```

```python
for i in range(columns):
    print(str(i)+". :", df.iloc[:,i].dtype)
```

```python
lst = []
col = df.columns
print(col)
for i in col:
    if i!="State":
        lst.append(i)
print(lst)
```

```python
df[lst] = df[lst].apply(pd.to_numeric)
```

```python
for i in range(columns):
    print(str(i)+". :", df.iloc[:,i].dtype)
```

```python
plt.figure(figsize=(30,10))
sns.heatmap(df.corr(),annot=True,linewidth=1)
```

```python
for i in lst:
    plt.xlabel(i)
    plt.ylabel("State")
    plt.title(i+" vs State")
    plt.figure(figsize=(30,10))
    ax = sns.barplot(x="State",y = i,data=df)
    ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
    #ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
    #plt.tight_layout()
    plt.show()
```
