```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from catboost import CatBoostRegressor
from sklearn.model_selection import train_test_split 
```

```python
laptops = '../input/laptop-specification-and-price-prediction/laptop_price.csv'

df = pd.read_csv(laptops)
```

```python
df.head()
```

```python
df['Price'] = df['Price'].str.replace(',', '')
df['Price'] = df['Price'].str.replace('₹', '')
```

```python
variaveis = ['Name','Processor','RAM','Operating System','Storage', 'Display','Warranty','rating']

x = df[variaveis]
y = df['Price']
```

```python
x.head()
```

```python
x['Name'].value_counts()
```

```python
x['RAM'].value_counts()
```

```python
plt.figure(figsize=(24, 10))
sns.countplot(x['RAM'])
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=0)
```

```python
categorical_features_indices = np.where(x.dtypes != np.float)[0]
```

```python
model = CatBoostRegressor(iterations=70, learning_rate=0.1, depth=3)
```

```python
model.fit(x_train, y_train, cat_features=categorical_features_indices, eval_set=(x_test, y_test), plot=True)
```
