```python
# Select only the important features i.e. the date and price
data = df_gold[["Date","Value"]] # select Date and Price
# Rename the features: These names are NEEDED for the model fitting
data = data.rename(columns = {"Date":"ds","Value":"y"}) #renaming the columns of the dataset
data.head(5)
```

```python
data.describe()
```

```python
# FB Prophet Model for predictionsfrom fbprophet import Prophet
from fbprophet import Prophet

m = Prophet(daily_seasonality = True) # the Prophet class (model)
m.fit(data) # fit the model using all data
```

```python
future_price = m.make_future_dataframe(periods=365) #we need to specify the number of days in future
prediction = m.predict(future_price)
m.plot(prediction)
plt.title("Prediction of the Gold Price using the FB Prophet")
plt.xlabel("Date")
plt.ylabel("Value")
plt.show()
```

```python
#### Improving the Model
m_2 = Prophet(n_changepoints=20, yearly_seasonality=True)
m_2.fit(data) # fit the model using all data
```

```python
future_price = m_2.make_future_dataframe(periods=365) #we need to specify the number of days in future
prediction = m_2.predict(future_price)
m.plot(prediction)
plt.title("Prediction of the Gold Price using the FB Prophet")
plt.xlabel("Date")
plt.ylabel("Value")
plt.show()
```
