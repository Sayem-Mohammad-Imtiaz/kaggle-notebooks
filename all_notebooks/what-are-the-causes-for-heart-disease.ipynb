```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'/kaggle/input/heart-disease-uci/heart.csv')
```

# Attribute Information

1. age
2. sex
3. chest pain type (4 values)
4. resting blood pressure
5. serum cholestoral in mg/dl
6. fasting blood sugar > 120 mg/dl
7. resting electrocardiographic results (values 0,1,2)
8. maximum heart rate achieved
9. exercise induced angina(Angina is a type of pain that occurs when not enough blood flows to the heart muscle.)
10. oldpeak = ST depression induced by exercise relative to rest
11. the slope of the peak exercise ST segment
12. number of major vessels (0-3) colored by flourosopy
13. thal(A blood disorder involving lower-than-normal amounts of an oxygen-carrying protein.): 3 = normal; 6 = fixed defect; 7 = reversable defect

```python
df.rename(columns={"cp": "chest_pain_type", "trestbps": "resting_bp", "restecg": "fasting_blood_sugar",
                  "thalach": "max_heart_rate", " exang":"exercise_reduced_angina"},inplace=True)
```

```python
df.head()
```

```python
df['target'].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df.corr(),annot=True)
```

we can see that cp(chest pain), thalach(max heart rate achieved), slope are highly correlated with target i.e higher these values higher the chances of getting the heart disease

```python
plt.figure(figsize=(15,6))
plt.title('5 Point summary of variables')
df.boxplot()
plt.xticks(rotation=45)
plt.show()
```

outlier alert: We can see that cholestrol has some very high values wrt normal(120) and max_heart_rate column has some lower values let's visualize them seperately

```python
plt.title('5 Point summary of cholestrol')
df['chol'].plot(kind='box',figsize=(6,8))
plt.yticks(range(200,600,50))
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['age'],df['chol'])
plt.title('Age vs Cholestrol')
plt.xlabel('Age')
plt.ylabel('Cholestrol Level')
```

we can see that with age cholestrol also increase but the value greater than 500 is still irrelevant(outlier)

```python
df = df[df['chol'] < 500]
```

```python
#verifying outliers
plt.title('5 Point summary of cholestrol')
df['chol'].plot(kind='box',figsize=(6,8))
plt.yticks(range(200,600,50))
plt.grid()
plt.show()
```

```python
plt.title('5 Point summary of max heart rate')
df['max_heart_rate'].plot(kind='box',figsize=(6,8))
plt.grid()
plt.show()
```

```python
df = df[df['max_heart_rate'] > 80]
```

```python
#verifying outliers
plt.title('5 Point summary of max heart rate')
df['max_heart_rate'].plot(kind='box',figsize=(6,8))
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15,6))
plt.title('Relation of age with heart disease')
sns.barplot(x=df['age'],y=df['target'])
```

We can see that there are higher chances of getting a heart disease at early 30's, early 40's and early 70's

```python
plt.figure(figsize=(10,6))
plt.scatter(df['age'],df['max_heart_rate'])
plt.title('Age vs heart rate')
plt.xlabel('Age')
plt.ylabel('Heart rate')
```

Heart rate decreases with age 

```python
plt.figure(figsize=(10,6))
sns.countplot(x=df['target'],hue=df['sex'],palette='Blues')
```

Comparatively higher females suffer from heart disease
