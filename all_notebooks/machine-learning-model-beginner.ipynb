```python
data = pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
data
```

```python
x = data.drop(columns='Outcome').values
x
```

```python
y = data['Outcome'].values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.218, train_size=0.7813, random_state=10, shuffle=False, stratify=None)

X_train
X_test
y_train
y_test
```

Modell trainieren

```python
# TODO: DecisionTreeClassifier aus Scikit-Learn importieren
from sklearn.tree import DecisionTreeClassifier

# TODO: Modell instanziieren
my_model = DecisionTreeClassifier()

# TODO: DecisionTreeClassifier auf den Daten trainieren

# Fit
# - X muss ein array-like mit 2 Achsen (Zeilen und Spalten) sein
# - y muss ein array-like mit 1 Achse sein, quasi ein Vektor,
#   gleiche LÃ¤nge wie die Anzahl der Zeilen in X
my_model.fit(X_train, y_train)
```

Modell evaluieren

```python
y_pred = my_model.predict(X_test)
y_pred
```

```python
accuracy = (y_test==y_pred).sum()/len(y_pred)
accuracy
```
