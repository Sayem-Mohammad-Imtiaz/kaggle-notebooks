**3.Apple**

```python
import matplotlib.pyplot as plt
import datetime

data = pd.read_csv('/kaggle/input/apple-aapl-historical-stock-data/HistoricalQuotes.csv')
data.head()
```

```python
data.isna().any()
```

```python
data.info()
```

```python
data = data.rename(columns={' Close/Last':'Close', ' Volume':'Volume', ' Open': 'Open', ' High':'High', ' Low':'Low'})
data['Close'] = data['Close'].str.replace('$', '').astype('float')
data['Open'] = data['Open'].str.replace('$', '').astype('float')
data['High'] = data['High'].str.replace('$', '').astype('float')
data['Low'] = data['Low'].str.replace('$', '').astype('float')
data.head()
```

```python
data['Open'].plot(figsize=(16, 6))
data.rolling(100).mean()['Open'].plot()
```

```python
training_df = data['Open']
training_df = pd.DataFrame(training_df)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0,1))
training_df_scaled = sc.fit_transform(training_df)
```

```python
X_train = []
y_train = []
for i in range(60, 2477):
    X_train.append(training_df_scaled[i-60:i, 0])
    y_train.append(training_df_scaled[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)


X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_train
```
