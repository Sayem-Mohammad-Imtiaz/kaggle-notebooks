I use Wilcoxon Matched Pairs Signed Rank Test to test the old price and discount price. It is the nonparametric equivalent of the parametric paired t-test. The Wilcoxon Signed Rank Test is more powerful than the sign test. This statistic differs from the sign test in that it considers the magnitude of the difference while the sign test does not. It uses more information from the sets of scores than the simple sign test.

from = https://www.sciencedirect.com/topics/medicine-and-dentistry/wilcoxon-signed-ranks-test

But, before that, I do the Kolmogorov-Smirnov Goodness of Fit Test to see normality in the data.

# Kolmogorov-Smirnov Goodness of Fit Test (K-S test)

The Kolmogorov-Smirnov Goodness of Fit Test (K-S test) compares data with a known distribution. Although the test is nonparametric — it doesn’t assume any particular underlying distribution — it is commonly used as a test for normality to see if the data is normally distributed.

from = https://www.statisticshowto.com/kolmogorov-smirnov-test/

I count the difference between discount price and old price to test the normality and the hypothesis. The old price is a real price and the discount price is a current price after getting a discount. The 'gap' group data is the difference between those two group price.

```python
data['gap'] = data['old_price'] - data['discount_price']
```

```python
sns.distplot(data['gap'])
plt.title('Difference Between Old Price And Discount Price')
plt.show()
```

```python
from scipy.stats import kstest
```

The hypothesis is:
# H0: Normal Distributed
# H1: Not Normal Distributed

I decide to use the group of difference called the gap.

```python
stats, pval = kstest(data['gap'], 'norm')
```

```python
stats
```

```python
pval
```

```python
pval < 0.05
```

In this test, the p-value is not lower than 0.05, so the H0 hypothesis is declined. It means that the distribution is not normal.

# Wilcoxon Matched Pairs Signed Rank Test

```python
from scipy.stats import wilcoxon
```

The hypothesis is:

# H0: P0 = P1
# H1: P0 ≠ P1

H0 means there is no difference between old price with discount price and H1 means there is a difference between old price with discount price.

The test is one-sided using 'alternative'.

```python
rank, pval = wilcoxon(data['gap'], alternative = 'less')
```

```python
rank
```

```python
pval
```

```python
pval < 0.05
```

In this test, the p-value is greater than 0.05, so the H0 hypothesis is accepted. It means that there is no difference between old with discount price.

# Spearman Rank Test

Spearman's rank-order correlation is the nonparametric test. Spearman's correlation coefficient measures the strength and direction of the association between two ranked variables.

from = https://statistics.laerd.com/statistical-guides/spearmans-rank-order-correlation-statistical-guide.php

```python
data[['old_price', 'discount_price']].corr(method = 'spearman')
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(data[['old_price', 'discount_price']].corr(method = 'spearman'), annot=True)
plt.title('Heatmap Correlation Between')
plt.show()
```

```python
sns.scatterplot(data['old_price'],data['discount_price'])
plt.title('Correlation Spearman')
plt.show()
```

The plot of old price with discount price, indicating a monotonic relationship. Monotonicity is less restrictive than that of a linear relationship. The scatter describe there is a correlation between those two variables.

# Conclusion:
* The data is not normally distributed based on The Kolmogorov-Smirnov Goodness of Fit Test (K-S test). The hypothesis is declined.
* Based on Wilcoxon Matched Pairs Signed Rank Test using the gap group data (the difference between discount price and old price), the hypothesis is accepted. It means that there is no meaningful difference at all between discount price and old price, the discount price is not lower than given.
* Based on Spearman Rank Test, there is a correlation between old price and discount price as seen on the scatter plot called monotonic relationship. In a monotonic relationship, the variables tend to move in the same relative direction, but not necessarily at a constant rate.
