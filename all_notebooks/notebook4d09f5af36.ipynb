```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
data_filepath = "../input/spotify-dataset-19212020-160k-tracks/data.csv"
spotify_data = pd.read_csv(data_filepath)
print("Data loaded Successfully")
```

```python
spotify_data.head()
```

```python
spotify_head = spotify_data.iloc[:20]

data_test = spotify_head.groupby(["artists"]).\
    agg({'popularity': 'mean', 'key': 'mean'}).\
        reset_index()

plt.figure(figsize=(20,10))
plt.title("Histogramme des popularités des 20 premiers artistes triès dans le data")


sns.barplot(x=data_test.popularity, y=data_test.artists)

plt.legend()
```

```python
spotify_data_2010 = spotify_data.loc[spotify_data.year == 2010]
data_test = spotify_data_2010.groupby(["artists"]).\
  agg({'popularity': 'max', 'danceability': 'max'}).\
        reset_index()

plt.figure(figsize=(20,10))
plt.title("Histogrammes des popularity et danceabilty des artistes des années 2010")

sns.distplot(a=data_test['popularity'],label="Popularité", kde=False)

plt.legend()


```
