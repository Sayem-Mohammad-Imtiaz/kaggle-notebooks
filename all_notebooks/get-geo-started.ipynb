```python
import numpy as np
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

ds=pd.read_csv("../input/forbes-billionaires-2021-30/forbes_billionaires_geo.csv")

from shapely import wkt

ds['geometry'] = ds['geometry'].apply(wkt.loads)

world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
ds = gpd.GeoDataFrame(ds).set_geometry('geometry')

base = world.plot(color='white', edgecolor='black', figsize=(16, 9))

ds.plot(ax=base, marker='o', color='magenta', markersize=1);
```
