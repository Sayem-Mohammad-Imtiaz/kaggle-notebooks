```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
X[:, 3] = labelencoder.fit_transform(X[:, 3])
ct =ColumnTransformer([('encoder', OneHotEncoder(), [3])],remainder='passthrough')
X= np.array(ct.fit_transform(X), dtype=np.float)
```

### Dummy Variable Trap

The Dummy Variable trap is a scenario in which the independent variables are multicollinear - a scenario in which two or more variables are highly correlated; in simple terms one variable can be predicted from the others.

The solution to the dummy variable trap is to drop one of the categorical variables (or alternatively, drop the intercept constant) - if there are m number of categories, use m-1 in the model, the value left out can be thought of as the reference value and the fit values of the remaining categories represent the change from this reference.

```python
# Avoiding Dummmy variable Trap
X = X[:, 1:]
```

Splitting Dataset Into Two Halves Training and Testing Part

```python
# Splitting the Dataset into Training and Testing Part
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Fitting Our Training part into Linear Regression Model

```python
# Fitting in Linear Regression Model
from sklearn.linear_model import LinearRegression
le = LinearRegression()
le.fit(X_train, y_train)
```

Predicting the Profit Value For X_test Data

```python
# Predicting the Data
y_pred_test = le.predict(X_test)
```

## Backward Elimination

Backward Elimination is a process of Selecting a Features from a dataset. This makes the model more precise. Selecting features for your model is the major task. 

### Steps to Implement Backward Elimination

![](http://3.bp.blogspot.com/-Q5NbRCOwVQc/WVCRFGON1QI/AAAAAAAABsc/Nzu8UtbejwoqFf6FYWhUGuWpMd-jDfAtwCLcBGAs/s1600/Capture5.PNG)

```python
import statsmodels.regression.linear_model as sm

# Adding a Constant bo in multilinaer regression feature equation
X = np.append(arr = np.ones((50,1)).astype(int), values = X, axis = 1)

# Writting all the indexes and copying all the features 
X_opt = X[:, [0, 1, 2, 3, 4, 5]]

# Using Signiface Level SL = 0.500

# Fit the Full model with all possible predictors
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()

#Seeing Summary of OLS fitting model
regressor_OLS.summary()
```

```python
# Removing b2 its p-value > SL
# Writting all the indexes and copying all the features 
X_opt = X[:, [0, 1, 3, 4, 5]]
# Fit the Full model with all possible predictors
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
# Seeing Summary of OLS fitting model
regressor_OLS.summary()
```

```python
# Removing b1 its p-value > SL
# Writting all the indexes and copying all the features 
X_opt = X[:, [0, 3, 4, 5]]
# Fit the Full model with all possible predictors
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
# Seeing Summary of OLS fitting model
regressor_OLS.summary()

```

```python
# Removing b4 its p-value > SL
# Writting all the indexes and copying all the features 
X_opt = X[:, [0, 3, 5]]
# Fit the Full model with all possible predictors
regressor_OLS = sm.OLS(endog = y, exog = X_opt).fit()
# Seeing Summary of OLS fitting model
regressor_OLS.summary()
```

Re-Creating Linear Model Now only for Specified features

```python
# Performing Linear Regression again by using main features only

# Allocating Data
X_BE, y_BE = X_opt, y

# Splitting the Dataset into Training and Testing Part
from sklearn.model_selection import train_test_split
X_BE_train, X_BE_test, y_BE_train, y_BE_test = train_test_split(X_BE, y_BE, test_size = 0.2, random_state = 0)

# Fitting in Linear Regression Model
le.fit(X_BE_train, y_BE_train)
```

```python
# Predicting the Data
y_BE_pred_test = le.predict(X_BE_test)
```
