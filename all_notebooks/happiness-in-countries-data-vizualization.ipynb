```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objs as go
```

```python
df2020=pd.read_csv('../input/world-happiness-report-2020/WHR20_DataForFigure2.1.csv')
df2021=pd.read_csv('../input/world-happiness-report-2021/world-happiness-report-2021.csv')
df_all_years=pd.read_csv('../input/world-happiness-report-2021/world-happiness-report.csv')
pd.set_option('max_rows',None)
```

```python
df2020=df2020.rename(columns={'Ladder score': 'Life Ladder','Healthy life expectancy':'Healthy life expectancy at birth'
                        ,'Logged GDP per capita':'Log GDP per capita'})
df2021=df2021.rename(columns={'Ladder score': 'Life Ladder','Healthy life expectancy':'Healthy life expectancy at birth'
                        ,'Logged GDP per capita':'Log GDP per capita'})
year2020=[2020 for number in range(len(df2020))]
df2020['year']=year2020
year2021=[2021 for number in range(len(df2021))]
df2021['year']=year2021
df_all_years=df_all_years[df_all_years['year']!=2020]
df2020=df2020.sort_values('Country name').reset_index()
df2021=df2021.sort_values('Country name').reset_index()
df_2020=df2020[['Country name','year','Regional indicator','Life Ladder','Log GDP per capita','Social support',
                'Healthy life expectancy at birth', 'Freedom to make life choices','Generosity','Perceptions of corruption']]
df_2021=df2021[['Country name','year','Regional indicator','Life Ladder','Log GDP per capita','Social support',
                'Healthy life expectancy at birth', 'Freedom to make life choices','Generosity','Perceptions of corruption']]
```

```python
df_all_years=df_all_years.merge(df_2020[['Country name','Regional indicator']],left_on='Country name',right_on='Country name')
df_all_years=df_all_years.append(df_2020)
df_all_years=df_all_years.append(df_2021)
```

```python
df_all_years=df_all_years.reset_index().drop(columns=['index'])
```

```python
df=pd.read_csv('../input/iso-codes/iso countries list (csv)')
df=df[['name','iso3']]
df.columns=['Country name','ISO CODE']
df_2020=df_2020.merge(df,left_on='Country name',right_on='Country name')
df_2021=df_2021.merge(df,left_on='Country name',right_on='Country name')
df_all_years=df_all_years.merge(df,left_on='Country name',right_on='Country name')
```
