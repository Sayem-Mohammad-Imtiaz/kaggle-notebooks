## How get the total medals by country?

```python
groupby_columns = ['medal_date', 'discipline_code', 'event', 'medal_code', 'medal_type', 'country_code']
df_medals_true = df_medals.groupby(groupby_columns)['athlete_name'].apply(list).reset_index()
print(df_medals_true.shape)
df_medals_true.head(10)
```

```python
df_medals_gr = df_medals_true.groupby(['country_code', 'medal_type'])['medal_code'].count().reset_index()
df_medals_gr.head(10)
```

```python
df_medals_country = df_medals_gr.pivot('country_code', 'medal_type', 'medal_code').fillna(0).astype(int)
df_medals_country['Total'] = df_medals_country.sum(axis=1).astype(int)
df_medals_country = df_medals_country.reset_index().sort_values(['Gold Medal', 'Silver Medal', 'Bronze Medal'], 
                                                                ascending=[False, False, False])
df_medals_country.columns.name='index'
df_medals_country = df_medals_country.rename(columns={'country_code': 'Country Code'})
df_medals_country['Rank'] = list(range(1, df_medals_country.shape[0]+1))
df_medals_country = df_medals_country[['Rank', 'Country Code', 'Gold Medal', 'Silver Medal', 'Bronze Medal', 'Total']]
df_medals_country = df_medals_country.reset_index(drop=True)
df_medals_country
```
