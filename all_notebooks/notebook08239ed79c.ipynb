```python
data=pd.read_csv('/kaggle/input/hotel-booking-demand/hotel_bookings.csv')
data.head()
```

```python
data.info()   #Properties of dataset
```

```python
data.describe()
```

```python
data.isnull().any()      #Check for any null values
```

```python
data.isnull().sum()     #Number of null values in each column
```

```python
data['children'].fillna(0,inplace=True)     #Replace null with 0
```

```python
data['children'].isnull().sum()        #Checking if all values are replaced
```

```python
data['country'].ffill(axis=0,inplace=True)         #Replace null with last value of column
```

```python
data['country'].isnull().sum()                   #Checking if all values are replaced
```

```python
data['agent'].fillna(data['agent'].mean(),inplace=True)           #Checking null with mean of a column
```

```python
data['agent'].isnull().sum()                    #Checking if all values are replaced
```
