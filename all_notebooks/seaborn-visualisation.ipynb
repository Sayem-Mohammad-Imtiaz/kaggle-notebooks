## Step 1: Load the data

A waiter recorded information about each tip he received over a period of a few months working in a restaurant. In all he recorded 244 tips.

Analyzing which factors might be impact the tip amount

```python
# Fill in the line below: Specify the path of the CSV file to read
my_filepath = '../input/tipping/tips.csv'

# Fill in the line below: Read the file into a variable my_data

tips = pd.read_csv(my_filepath,parse_dates=True)
tips.head()
```

## Is there a relationship between bill amount and tip amount ?

```python
plt.figure(figsize=(12,6))
plt.title("Tip Amount Per total of bill in Dollars ")


sns.regplot(x=tips['total_bill'], y=tips['tip'])


plt.xlabel('bill amount (dollars)')
plt.ylabel('tip amount (dollars)')
```

```python
       
sns.jointplot(x= 'total_bill', y= 'tip', data = tips, kind="hex")
```

**Insight:** The scatter plot shows a strong correlation between the two variables. this tells us that bill value plays a strong role in the amount of the tip.

The jointplot shows the most amount of bills are between 10 to 18 dollars and tips are between 2 to 3 dollars.

## Distribution of the Party size, tip amount and bill amount !!

```python
plt.title('Distribution of Tip Amount and Party Size')
tips1 = tips[['tip', 'size']]
sns.boxplot(data= tips1)
```

```python
print(tips['tip'].mean())
print(tips['size'].mean())
```

```python
plt.title('Distribution of Bill Amount')

sns.boxplot(x= tips['total_bill'], orient='v')
```

```python
tips['total_bill'].mean()
```

```python
plt.figure(figsize=(12,6))

plt.title('Distribution of Bill and Tip Amount in Dollars')
sns.distplot(a = tips['tip'], bins= 20, label = "tip in dollars")
sns.distplot(a = tips['total_bill'], bins= 20, label = "bill in dollars", color ="purple")
plt.legend()
```

**Insight:** The box chart shows that most party size where has been got tips are about between 2 to 3 people. Most amount of tip is about between 2 to 4 dollars. For bill amount is about between 15 to 25 dollars

## Which day of the week, has been got more tips ? 

```python
plt.figure(figsize=(8,6))
plt.title("Tips Per Day of Week")


sns.countplot(x=tips['day'],  color = 'purple')
```

**Insight:** Based on the data, it sounds Saturday is more luckey day, as has been got tip more than 80 times. unlikely Friday, it is the least day in getting tips

## Which meal has more the tip amount ? 

```python
plt.figure(figsize=(8,6))
plt.title("Tips Amount Per Type of Meal and Gender ")

sns.countplot(x="time", hue="sex", data=tips)

plt.xlabel("meal type")
plt.ylabel("Tip count")
plt.legend()
```

**Insights:** Male in dinner meal are more likely give tip 

```python
plt.figure(figsize=(8,6))
plt.title("Tips Amount Per Type of Meal and Gender ")

sns.countplot(x="time", hue="sex", data=tips)

plt.xlabel("meal type")
plt.ylabel("Tip count")
plt.legend()
```

```python
plt.figure(figsize=(12,8))

plt.title("Amount of Bill per Day of week and smoker or not  ")

sns.boxplot(x = "day", y = "total_bill", hue="smoker", data=tips)

plt.xlabel("day of week")
plt.ylabel("bill (dollars)")
```
