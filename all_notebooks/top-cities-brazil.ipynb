```python
df=pd.read_csv("../input/top-cities-brazil/topCidadesBrasil.csv")
df.head()
```

#Codes by Taikutsu https://www.kaggle.com/ravels1991/simple-eda/comments

```python
plt.figure(figsize=(20,12))
g = sns.scatterplot(x='lng', y='lat', data=df, hue='capital')
g.legend(loc='center left', bbox_to_anchor=(1.25, 0.5), ncol=1);
```

```python
g = sns.lmplot(x="lng", y="lat", data=df,
           fit_reg=False, scatter_kws={"s": 30}, hue='country', height=10)
plt.title('Top Cities Brazil')
plt.show()
```

```python
#https://www.kaggle.com/olgabelitskaya/parts-of-speech
plt.figure(figsize=(10,5))
sns.countplot(y="capital",data=df,
             facecolor=(0,0,0,0),linewidth=5,
             edgecolor=sns.color_palette("winter"))
plt.title('Top Cities Brazil',
         fontsize=15);
```

#Codes from Iván Hernández Cazorla https://www.kaggle.com/ivanhercaz/museums-in-tenerife-location-map

```python
fig_px = px.scatter_mapbox(df, lat="lat", lon="lng",
                           hover_name="city",
                           zoom=11, height=300)
fig_px.update_layout(mapbox_style="open-street-map",
                     margin={"r":0,"t":0,"l":0,"b":0})

fig_px.show()
```

```python
fig_px.update_traces(marker={"size": [10 for x in df]})
```
