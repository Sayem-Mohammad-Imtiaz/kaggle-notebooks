```python
X=df1[[ 'Adult_Mortality','GDP','Income_composition_of_resources','Schooling']]
```

Determining the variables used in our model in link with the correlation with life expectancy.

```python
X.corr()
```

```python
Y=df1['Life_expectancy']
```

```python
Y
```

```python
X
```

```python
Y.fillna(Y.mean())
```

Testing the linear regression for our model

```python
result=train_test_split(X,Y,test_size=0.02)
```

```python
X_train = result[0] 
X_test  = result[1]
Y_train = result[2]
Y_test  = result[3]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
X_train.head(15)
```

```python
Y_train.head(15)
```

Fitting our model to the data used

```python
Y_train
```

```python
Y_train.fillna(Y_train.mean())
```

```python
lr.fit(X_train,Y_train)
```

```python
Ypred=lr.predict(X_test)
```

```python
Ypred-Y_test
```

```python
np.linalg.norm(Y_test-Ypred)/np.linalg.norm(Y_test)
```

The relative error is of 6.5% which is quit good! Let's try by droping or adding another variable

Deleting Adult mortality of our model as the corr√©lation is lower than the others with life expectancy

```python
X2=df1[['GDP','Income_composition_of_resources','Schooling']]
```

```python
Y2=df1['Life_expectancy']
```

```python
result=train_test_split(X2,Y2,test_size=0.02)
```

```python
X2_train = result[0] 
X2_test  = result[1]
Y2_train = result[2]
Y2_test  = result[3]
```

```python
Y2pred=lr.predict(X2_test)
```

```python
np.linalg.norm(Y2_test-Y2pred)/np.linalg.norm(Y2_test)
```
