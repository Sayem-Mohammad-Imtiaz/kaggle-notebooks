# 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<a id = '2'> </a> <br>
# 2. Load the Data

```python
df = pd.read_csv('/kaggle/input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
df
```

```python
df.info()
```

All data is numeric. That's why there is no need to encode.

<a id = '3'> </a> <br>
# 3. Taking Care of Missing Data
There are no missing values in this data as you can see below.

```python
df.notnull().all()
```

## Set Independent and Dependent Variable

```python
#Independent Variable
x = df.iloc[:, :-1].values
#Dependent Variable
y = df.iloc[:, -1].values
y = y.reshape(len(y), 1)
```

<a id = '4' > </a> <br>
# 4. Split the Dataset into Training Set and Test Set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 19)
```

<a id = '5'> </a> <br>
# 5. Feature Scaling
I will use standardization.


```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()
x_train = sc_x.fit_transform(x_train)
y_train = sc_y.fit_transform(y_train)
```

<a id = '6'> </a> <br>
# 6. Training the Support Vector Regression Model on the Training Set

```python
from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(x_train, y_train)
```

<a id = '7'> </a> <br>
# 7. Predicting the Test Set Results

```python
y_pred = sc_y.inverse_transform(regressor.predict(sc_x.transform(x_test)))
```

<a id = '8'> </a> <br>
# 8. Calculate the Accuracy

Here I will use R squared for calculating the accuracy.

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
