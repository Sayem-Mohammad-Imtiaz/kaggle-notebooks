# **Data cleaning and wrangling**
* Convert object to int for
*  * rank
* * discipline
* * sex

```python
# Convert rank to int
rank = data.groupby("rank").mean()
rank
```

```python
rank_dict = {"Prof": 3, "AssocProf" : 2, "AsstProf" : 1}
updated_data = data.copy()
updated_data["rank"] = updated_data["rank"].map(rank_dict)
updated_rank = updated_data.groupby("rank").mean()
updated_rank
```

```python
# Convert discipline to int
discipline = data.groupby("discipline").mean()
discipline
```

```python
discipline_dict = {"A": 1, "B" : 2}
updated_data["discipline"] = updated_data["discipline"].map(discipline_dict)
updated_discipline = updated_data.groupby("discipline").mean()
updated_discipline
```

```python
# Convert sex to int
sex = data.groupby("sex").mean()
sex
```

```python
sex_dict = {"Female": 1, "Male" : 2}
updated_data["sex"] = updated_data["sex"].map(sex_dict)
updated_sex = updated_data.groupby("sex").mean()
updated_sex
```

```python
updated_data.info()
```

# **Store features as X**

```python
updated_data.columns
```

```python
X = updated_data[['rank','discipline','yrs.since.phd','yrs.service','sex']]
X
```

# **Store target as y**

```python
y = updated_data[["salary"]]
y
```

# **Train, test and split data**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 80)
```

# **Fit data: Linear regression**

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_prediction = lr.predict(X_test)
y_prediction[:5]
```

```python
y_test.describe()
```

```python
# Calculate the root mean sqaured error
RMSE = sqrt(mean_squared_error(y_true = y_test, y_pred = y_prediction))
RMSE
```

```python
r_squared = lr.score(X_test,y_test)
r_squared
```
