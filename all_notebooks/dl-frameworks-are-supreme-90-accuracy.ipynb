```python
from fastai.tabular.all import *
```

```python
dls = TabularDataLoaders.from_csv('/kaggle/input/heart-disease-prediction/Heart_Disease_Prediction.csv', 
                                  path='/kaggle/input/heart-disease-prediction/', y_names="Heart Disease",
    cat_names = ['Sex','Chest pain type','FBS over 120','EKG results','Exercise angina','Slope of ST','Number of vessels fluro','Thallium'],
    cont_names = ['Cholesterol','BP','Age','Max HR','ST depression'],
    procs = [Categorify, FillMissing, Normalize],bs=32)
```

```python
learn = tabular_learner(dls, metrics=accuracy)

learn.fit_one_cycle(15)
```
