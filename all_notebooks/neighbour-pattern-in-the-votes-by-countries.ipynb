# Here we will find the longitude and lattitude of Every countries in our list to make map visuals

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="your_app_name")
def locate(country):
    try:
        # Geolocate the center of the country
        loc = geolocator.geocode(country)
        # return latitude and longitude
        return (loc.latitude, loc.longitude)
    except:
        return np.nan
countriesloc={}
for x in df.iloc[:,0].values:
    countriesloc[x]=locate(x)
```

> **Let's set our voters as a index for simplicity**

```python
df.set_index('Country (Voters (vertical), Finalists (horizontal))',inplace=True)
df.head(3)
```

# Let's Visualise The Each Voter's Votes Country Wise.
> **Some Points to be noted here are :**
* We Show all those countries in a map to whom Voter Voted greater Than 0
* Red marker is the Voter and blue markers denote the countries to whom it Votes.
* Each blue Marker is the Pop up which shows the point from 1-12.

```python
def votes_to(country):
    mp=folium.Map(countriesloc[country],zoom_start=3)
    folium.Marker(countriesloc[country],icon=folium.Icon(color="red")).add_to(mp)
    lis=[(x,y) for x,y in zip(df.loc[country,:].index,df.loc[country,:].values)]
    for x in lis:
        if(x[1]>0):
            folium.Marker(countriesloc[x[0]], popup=x[1]).add_to(mp)
    display(mp)
for x in df.index:
    print(x)
    votes_to(x)
```

# From The Visualizations These Points Can Be Concluded
> **Generally Contries Support Their Neibours i.e They Generally Vote For Their Neighbours**
