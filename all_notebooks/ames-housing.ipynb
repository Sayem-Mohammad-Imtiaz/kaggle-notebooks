```python
def missing_percent(df):
    nan_percent= 100*(df.isnull().sum()/len(df))
    nan_percent= nan_percent[nan_percent>0].sort_values()
    return nan_percent
```

```python
df_obj= pd.get_dummies(df_obj, drop_first=True)
```

```python
df_obj.shape
```

```python
Final_df= pd.concat([df_num, df_obj], axis=1)
```

```python
Final_df.head()
```

```python
X = Final_df.drop('SalePrice', axis=1)
y = Final_df['SalePrice']
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polyconverter = PolynomialFeatures(include_bias=False)
```

```python
polyfeatures = polyconverter.fit(X)
```

```python
polyfeatures = polyconverter.transform(X)
```

```python
polyfeatures.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(polyfeatures,y,test_size=0.3,random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
pmodel = LinearRegression()
```

```python
pmodel.fit(X_train,y_train)
```

```python
y_pred = pmodel.predict(X_test)
```

```python
Residuals = pd.DataFrame(y_test - y_pred)
Residuals.head()
```

```python
from sklearn import metrics
MAE= metrics.mean_absolute_error(y_test, y_pred)
MSE= metrics.mean_squared_error(y_test, y_pred)
RMSE= np.sqrt(MSE)
pd.DataFrame([MAE, MSE, RMSE], index=['MAE', 'MSE', 'RMSE'], columns=['Metrics'])
```
