```python
userdata.to_csv('csv_to_submit.csv', index = False)

df=userdata[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','ca']].copy()

df.corr(method ='kendall') 
```

```python
from sklearn.decomposition import PCA 
pca = PCA(n_components=3)
principalComponents = pca.fit_transform(df)
principalDf = pd.DataFrame(principalComponents, columns = ['fitness1','fitness2','fitness3'])

column = principalDf['fitness1']
max_value = column.max()
min_value = column.min()


print(max_value)
print(min_value)

column = principalDf['fitness2']
max_value = column.max()
min_value = column.min()


print(max_value)
print(min_value)

column = principalDf['fitness3']
max_value = column.max()
min_value = column.min()


print(max_value)
print(min_value)

print(principalDf)
```

```python
for i,j in principalDf.iterrows():
    j.fitness1 = int(math.floor(j.fitness1)) + 120
    j.fitness2 = int(math.floor(j.fitness2)) + 56
    j.fitness3 = int(math.floor(j.fitness3)) + 36
    
column = principalDf['fitness1']
max_value = column.max()
min_value = column.min()


print(max_value)
print(min_value)

column = principalDf['fitness2']
max_value = column.max()
min_value = column.min()


print(max_value)
print(min_value)

column = principalDf['fitness3']
max_value = column.max()
min_value = column.min()


print(max_value)
print(min_value)

print(principalDf)
```

```python
l = list(principalDf['fitness1'])
m = list(principalDf['fitness2'])
n = list(principalDf['fitness3'])

print(len(set(l)))
print(len(set(m)))
print(len(set(n)))

fitness1 = dict([(y,x+1) for x,y in enumerate(set(l))])
fitness2 = dict([(y,x+1) for x,y in enumerate(set(m))])
fitness3 = dict([(int(y),x+1) for x,y in enumerate(set(n))])



t = tf.zeros([437,136,100])
tensor = tensorly.tensor(t) 
print(tensor)
```

```python
target = userdata["target"]
principalDf = principalDf.join(target)
```

```python


tensor_array = np.zeros((437,136,100))
count = 0
for i,j in principalDf.iterrows():
    count = count + 1
    tensor_array[int(j.fitness1),int(j.fitness2),int(j.fitness3)] = j.target + 1

unique,counts = np.unique(tensor_array,return_counts=True)
print(dict(zip(unique,counts)))
print(count)

    
    
```

```python
tensor2 = tf.convert_to_tensor(tensor_array)
    
```

```python
factors = non_negative_parafac(tensor2,rank = 3, verbose = 2)
print(factors)
```

```python
print(len(factors))
regenerated_tensor = tensorly.kruskal_to_tensor(factors)
print(regenerated_tensor.shape)
```

```python
#Error measurement
print("Root Mean Square error = "+str(tensorly.metrics.regression.RMSE(tensor, regenerated_tensor, axis=None)))

```
