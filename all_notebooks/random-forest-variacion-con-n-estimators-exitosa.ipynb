```python
# importas modulos y paquetes 
import pandas as pd
from sklearn.ensemble import RandomForestRegressor 
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
#sacamos la base de datos a usar
directorio="/kaggle/input/house-prices-dataset/train.csv"
base_de_datos=pd.read_csv(directorio)
caracteristicas= ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X=base_de_datos[caracteristicas]
y=base_de_datos.SalePrice
```

```python
#partimos la data en secciones 
train_X,val_X,train_y,val_y=train_test_split(X,y,random_state=0)
```

```python
#modelo de prediccion
def modelo(estimacion):
    modelo_prediccion=RandomForestRegressor(n_estimators=estimacion,criterion="mae",random_state=1)
    modelo_prediccion.fit(train_X,train_y)
    return(modelo_prediccion)
```

```python
#predicciones terminadas 
#aca varias el n estimators y ajustas la prescicion del resultado  
for estimacion in range(1,50):
    modelo_variable=modelo(estimacion)
    prediccion=modelo_variable.predict(val_X)
    #error detectado 
    mae=mean_absolute_error(val_y,prediccion)
    print("con la estimacion:",estimacion,"\t el error es de :",mae)
```
