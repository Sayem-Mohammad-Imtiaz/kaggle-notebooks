```python
count=df.shape[0]
array1 = np.random.randint(low = 0, high = 3, size = count)
array2 = np.random.randint(low = 0, high = 3, size = count)
for i in Country_list:
    temp=df_dict[i].split('|||')
    t=l=0
    for j in temp:
        x = json.loads(j)
        #print(x)
        t+=float(x['Latitude'])
        l+=float(x['Longitude'])
    array1[i]=t/len(temp)  
    array2[i]=l/len(temp)
```

```python
dLat = df['avg_latitude']- 30.5928
dLon = df['avg_longitude'] - 114.3055
dLon=dLon*math.pi/180.0
dLat=dLat*math.pi/180.0
c = (np.sin(dLat/2)* np.sin(dLat/2)+ np.cos(np.deg2rad(30.5928))* np.cos(np.deg2rad(df['avg_latitude']))*np.sin(dLon/2)* np.sin(dLon/2))
d = 2 * np.arctan2(np.sqrt(c), np.sqrt(1-c)) *6373
```

```python
def get_cmap(n, name='hsv'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)
```
