```python
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.linear_model import LogisticRegression
df= pd.read_csv("../input/hr-analytics/HR_comma_sep.csv")
```

```python
pd.crosstab(df.Department,df.left).plot(kind='bar')
```

```python
from sklearn.preprocessing import StandardScaler,LabelEncoder
Label_encoder_depart=LabelEncoder()
Label_encoder_salary=LabelEncoder()
depart=Label_encoder_depart.fit_transform(df["Department"])
salary=Label_encoder_salary.fit_transform(df["salary"])
df["Department"]=depart
df["salary"]=salary

target=df["left"]
inputs=df.iloc[:,0:9].values
x_train,x_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
cm=confusion_matrix(y_test,predict)
cm
```

```python
model.score(x_train,y_train)
```
