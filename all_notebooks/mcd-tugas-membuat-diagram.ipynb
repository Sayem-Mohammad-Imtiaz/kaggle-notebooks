```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Import Data

```python
df = pd.read_csv('../input/kpopdb/kpop_idols_boy_groups.csv')
df.head(10)
```

```python
df.info()
```

# Line Chart

```python
# Dengan matplotlib
plt.plot(df['Name'], df['Members'], color='blue', marker='o')
plt.title('Name Vs Members', fontsize=14)
plt.xlabel('Name', fontsize=14)
plt.ylabel('Members.', fontsize=14)
plt.grid(True)
plt.show()
```

```python
# Dengan Seaborn
plot = sns.lineplot(x="Name", y="Members", data=df)
```

# Histogram

```python
# Dengan Matplotlib
plt.hist(df['Members'], bins=9, align='right', color='green', edgecolor='black')
```

```python
# Dengan Seaborn
# Membuat histogram dari kolom 'Members'
plot2 = sns.displot(data=df, x="Members", kde=True)
```

```python
plot3 = sns.scatterplot(x=df['Orig. Memb.'], y=df['Members'])
```

# Box Plot

```python
# ubah tipe data 'Active' menjadi kategori
df['Active'] = df['Active'].astype('category')
df.info()
```

```python
p2 = sns.boxplot(x="Members", y="Active", data=df)
```

# Styling

```python
# You can use figure sizing from Matplotlib 
plt.figure(figsize=(8,4)) 

# Change size of lables, lines and other elements to best fit 
# how you will present your data (paper, talk, poster) 
sns.set_context('paper', font_scale=1.4) 
sns.jointplot(x='Members', y='Orig. Memb.', data=df, kind='reg') 

# Get rid of spines 
# You can turn of specific spines with right=True, left=True 
# bottom=True, top=True 
sns.despine(left=False, bottom=False)
```
