```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from numpy import mean
from sklearn import metrics
import seaborn as ss
import os
```

```python
for dirname, _, filenames in os.walk('/kaggle/input/bsl-digits-dataset/BSL Digits Dataset/CSV Files/'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
sign0 = pd.read_csv("/kaggle/input/bsl-digits-dataset/BSL Digits Dataset/CSV Files/Output - 0.csv")
sign1 = pd.read_csv("/kaggle/input/bsl-digits-dataset/BSL Digits Dataset/CSV Files/Output - 1.csv")
sign2 = pd.read_csv("/kaggle/input/bsl-digits-dataset/BSL Digits Dataset/CSV Files/Output - 2.csv")
sign3 = pd.read_csv("/kaggle/input/bsl-digits-dataset/BSL Digits Dataset/CSV Files/Output - 3.csv")
sign4 = pd.read_csv("/kaggle/input/bsl-digits-dataset/BSL Digits Dataset/CSV Files/Output - 4.csv")
sign5 = pd.read_csv("/kaggle/input/bsl-digits-dataset/BSL Digits Dataset/CSV Files/Output - 5.csv")
sign6 = pd.read_csv("/kaggle/input/bsl-digits-dataset/BSL Digits Dataset/CSV Files/Output - 6.csv")
sign7 = pd.read_csv("/kaggle/input/bsl-digits-dataset/BSL Digits Dataset/CSV Files/Output - 7.csv")
sign8 = pd.read_csv("/kaggle/input/bsl-digits-dataset/BSL Digits Dataset/CSV Files/Output - 8.csv")
sign9 = pd.read_csv("/kaggle/input/bsl-digits-dataset/BSL Digits Dataset/CSV Files/Output - 9.csv")
```

```python
df = pd.concat([sign0, sign1, sign2, sign3, sign4, sign5, sign6, sign7, sign8, sign9])
```

```python
features = df.iloc[:,1:-1]
labels = df.iloc[:,-1]
```

```python
features_np = np.array(features)
```

```python
from sklearn.preprocessing import StandardScaler

sc_X = StandardScaler()
x = sc_X.fit_transform(features_np)
y = labels
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=0)
cv = KFold(n_splits=10, random_state=1, shuffle=True)
```

**KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
modelKNN = KNeighborsClassifier(n_neighbors=5)
modelKNN.fit(x_train, y_train)

y_predKNN = modelKNN.predict(x_test)
scores = cross_val_score(modelKNN, x_test, y_test, scoring='accuracy', cv=cv, n_jobs=-1)
accuracy_KNN = (mean(scores)*100)
print('Accuracy (KNN) : %.2f' % accuracy_KNN, '%')
```

**SVM - Linear**

```python
from sklearn.svm import SVC
modelSVM_Linear = SVC(kernel = 'linear')
modelSVM_Linear.fit(x_train, y_train)

y_predSVM_Linear = modelSVM_Linear.predict(x_test)
scores = cross_val_score(modelSVM_Linear, x_test, y_test, scoring='accuracy', cv=cv, n_jobs=-1)
accuracy_SVM_Linear = (mean(scores)*100)
print('Accuracy (SVM - Linear) : %.2f' % accuracy_SVM_Linear, '%')
```

**SVM - Polynomial**

```python
from sklearn.svm import SVC
modelSVM_Poly = SVC(kernel = 'poly', degree=3)
modelSVM_Poly.fit(x_train, y_train)

y_predSVM_Poly = modelSVM_Poly.predict(x_test)
scores = cross_val_score(modelSVM_Poly, x_test, y_test, scoring='accuracy', cv=cv, n_jobs=-1)
accuracy_SVM_Poly = (mean(scores)*100)
print('Accuracy (SVM - Polynomial) : %.2f' % accuracy_SVM_Poly, '%')
```

**SVM - RBF**

```python
from sklearn.svm import SVC
modelSVM_RBF = SVC(kernel = 'rbf', degree=5)
modelSVM_RBF.fit(x_train, y_train)

y_predSVM_RBF = modelSVM_RBF.predict(x_test)
scores = cross_val_score(modelSVM_RBF, x_test, y_test, scoring='accuracy', cv=cv, n_jobs=-1)
accuracy_SVM_RBF = (mean(scores)*100)
print('Accuracy (SVM - RBF) : %.2f' % accuracy_SVM_RBF, '%')
```

**XGB**

```python
from xgboost import XGBClassifier
modelXGB = XGBClassifier()
modelXGB.fit(x_train, y_train)

y_predXGB = modelXGB.predict(x_test)
scores = cross_val_score(modelXGB, x_test, y_test, scoring='accuracy', cv=cv, n_jobs=-1)
accuracy_XGB = (mean(scores)*100)
print('Accuracy (XGB) : %.2f' % accuracy_XGB, '%')
```

**RFC**

```python
from sklearn.ensemble import RandomForestClassifier
modelRFC = RandomForestClassifier(n_estimators = 10, criterion = 'gini')
modelRFC.fit(x_train, y_train)

y_predRFC = modelRFC.predict(x_test)
scores = cross_val_score(modelRFC, x_test, y_test, scoring='accuracy', cv=cv, n_jobs=-1)
accuracy_RFC = (mean(scores)*100)
print('Accuracy (RFC) : %.2f' % accuracy_RFC, '%')
```

**DTC**

```python
from sklearn.tree import DecisionTreeClassifier
modelDTC = DecisionTreeClassifier(criterion = 'gini')
modelDTC.fit(x_train, y_train)

y_predDTC = modelDTC.predict(x_test)
scores = cross_val_score(modelDTC, x_test, y_test, scoring='accuracy', cv=cv, n_jobs=-1)
accuracy_DTC = (mean(scores)*100)
print('Accuracy (DTC) : %.2f' % (mean(scores)*100), '%')
```

**Comparison**

```python
xAxis = ['XGB', 'KNN', 'SVM - L', 'SVM - P', 'SVM - R', 'DTC', 'RFC']
yAxis = [accuracy_XGB, accuracy_KNN, accuracy_SVM_Linear, accuracy_SVM_Poly, accuracy_SVM_RBF, accuracy_DTC, accuracy_RFC]

dframe = pd.DataFrame({"Classifiers" : xAxis, "Accuracy" : yAxis})


plt.figure(figsize=(16, 12))
splot = ss.barplot(x = "Classifiers",y = "Accuracy", data = dframe)
plt.xlabel("Classifiers", size=16)
plt.ylabel("Accuracy", size=16)
plt.savefig("BSL Sign Digits - Accuracy Comparison.png")
```
