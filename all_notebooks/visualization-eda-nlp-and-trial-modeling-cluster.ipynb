2. NLP

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation
import string
import squarify
import seaborn as sns

list_stopwords = set(stopwords.words('english'))
```

```python
df9=df[['ReasonsForLack']]
df9=df9.dropna()
```

```python
df9
```

```python
df9['ReasonsForLack'] = df9['ReasonsForLack'].str.lower()
df9['ReasonsForLack'] = df9['ReasonsForLack'].apply(word_tokenize)
df9['ReasonsForLack'] = df9['ReasonsForLack'].apply(lambda x: [word for word in x if word not in list_stopwords])
df9['ReasonsForLack'] = df9['ReasonsForLack'].apply(lambda x : [word.translate(str.maketrans('', '', string.punctuation)) for word in x])
df9['ReasonsForLack'] = df9['ReasonsForLack'].apply(lambda x : [word for word in x if len(word) > 1])
```

```python
df9
```

```python
df9_e=pd.DataFrame(df9['ReasonsForLack'].explode())
```

```python
df9_e
```

```python
df9_e=pd.DataFrame(df9_e.groupby('ReasonsForLack')['ReasonsForLack'].count().sort_values(ascending=False).head(50))
df9_e=df9_e.rename(columns={'ReasonsForLack': 'num'})

x = df9_e['num']
label = df9_e.index
squarify.plot(x, label=label,color=sns.color_palette('husl'))
plt.axis('off')
plt.show()
```

3.Trial Prediction Model by LGBM without tuning

```python
df1=df.drop('ReasonsForLack',axis=1)
```

```python
df1=pd.get_dummies(df1)
```

```python
df1
```

```python
import lightgbm as lgb
from sklearn.metrics import log_loss
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
X = df1.drop('y',axis=1)
Y = df1['y']
```

```python
pd.set_option('display.max_columns', 50)
```

```python
X
```

```python
X=X.rename(columns={'EducationSector_Art, Music or Design':'Edu_Art Music or Design',
                  'EducationSector_Economic Sciences, Business Studies, Commerce and Law':'Edu_Eco Business Commerce and Law',
                  'EducationSector_Medicine, Health Sciences':'Edu_Medi Health',
                  'EducationSector_Teaching Degree (e.g., B.Ed)':'Edu_Teaching'})
```

```python
train_X, test_X, train_Y, test_Y = train_test_split(X, Y,test_size=0.3,shuffle=True,random_state=0)
```

```python
lgb_train = lgb.Dataset(train_X, train_Y)
lgb_test = lgb.Dataset(test_X, test_Y, reference=lgb_train)
```

```python
params = {'objective':'binary', 
              'metric':'binary_logloss', 
              'verbosity': -1}
```

```python
lgb_results = {}
model = lgb.train(params=params,
                  train_set=lgb_train,
                  valid_sets=[lgb_train, lgb_test],
                  valid_names=['Train', 'Test'],
                  num_boost_round=100,
                  early_stopping_rounds=10,
                  evals_result=lgb_results)
```

By PCA, I will get 2 components for clusters.

```python
from sklearn.decomposition import PCA
```

```python
X1=X.drop('cluster',axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
clustering_sc = sc.fit_transform(X1)
```

```python
x1 = clustering_sc
pca = PCA(n_components=2)
pca.fit(x1)
x_pca = pca.transform(x1)
df13= pd.DataFrame(x_pca)
```
