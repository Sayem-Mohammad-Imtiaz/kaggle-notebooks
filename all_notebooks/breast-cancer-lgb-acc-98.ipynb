```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('../input/breast-cancer-wisconsin-data/data.csv')
```

```python
data.head(4)
```

```python
data.info()
```

```python
data.drop(['Unnamed: 32', 'id'], axis=1 , inplace= True)
```

```python
Label= data['diagnosis'].astype('str')
data.drop('diagnosis', axis=1, inplace= True)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encode=LabelEncoder()
Label =label_encode.fit_transform(Label)
```

<a id="1"></a>
<h2 style='background:gray; border:0; color:white'> <center> Data analysis and Visualization (EDA) </center></h2>

```python
plt.figure(figsize=(16, 16))
sns.heatmap(data.corr(), vmin=-1, vmax=1, annot= True )
```

```python
# check for balancity 
plt.title("The diagnosis of breast tissues (M = malignant, B = benign)")
sns.countplot(Label)
plt.xlabel("Severity")
```

<a id='2'> </a>
<h2 style='background:gray; border:0; color:white'> <center> Modelling and prediction </center></h2>

```python

from xgboost import XGBClassifier
from sklearn.metrics import f1_score, recall_score, accuracy_score
import xgboost as xgb
from lightgbm import LGBMClassifier
from numpy import mean
from numpy import std
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data,Label,random_state=0,test_size=0.2)
```

```python
model_lgb= LGBMClassifier()
model_lgb.fit(x_train, y_train)
```

<a id= '3'> </a>
<h2 style='background:gray; border:0; color:white'> <center> Model evaluation  </center> </h2>

```python
predictions= model_lgb.predict(x_test)
print(accuracy_score(y_test, predictions))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predictions))
```

```python
# checking the model performanec using cross validation
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=2, random_state=1)
n_scores = cross_val_score(model_lgb, data, Label, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```
