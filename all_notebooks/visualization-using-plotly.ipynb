# EDA ON MAHARASHTRA COVID CASES:

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

import plotly.express as px
import warnings
warnings.simplefilter("ignore")
```

```python
data = pd.read_csv("../input/latest-covid19-cases-maharashtra-india/Maharashtra Latest Covid Cases.csv")
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

No null values

# Visualizing Districts and the number of Positive Cases; their fatality rates:

```python
fig = px.bar(data.sort_values("Positive Cases"),x='Districts',y='Positive Cases',color = 'Fatality Rate (%)',template="simple_white")
fig.show()
```

INFERENCE: Pune has the highest number of positive cases but the fatality rate is 1.6%, so there are more cases and more people who tend to recover.

# Visualizing Districts and the number of Positive Cases; their Recovery rates:

```python
fig = px.scatter(data,x='Districts',y='Positive Cases',color = 'Recovery Rate (%)',template="ggplot2")
fig.show()
```

INFERENCE:  We can see that districts like Solapur, Sangli, Ratnagiri, Kohlapur have less number positive cases and the recovery rate is also comparatively less than that of other districts.

# Visualizing the number of Deceased people in various Districts:

```python
fig = px.sunburst(data, path=['Districts'], values='Deceased',
                  color_continuous_scale='RdBu')
fig.update_layout(
    margin = dict(t=10, l=10, r=10, b=10)
)
fig.show()
```

INFERENCE: The number of deceased is greater in Pune and Mumbai, where the cases are also high.

# Visualizing the number of people Recovered wrt the recovery rate:

```python
fig = px.bar_polar(data, r="Recovered", theta="Districts",
                   color="Recovery Rate (%)", template="plotly_dark",
                   color_discrete_sequence= px.colors.sequential.Plasma_r)

fig.update_layout(
    title='RECOVERY OF PATIENTS IN MAHARASHTRA',
    font_size=12,
    legend_font_size=12,
    polar_angularaxis_rotation=45,

)
fig.show()
```
