## 1. Calculate the Percentage change of returns
* use the function : pct_change

```python
df=pd.read_csv('../input/stocks-data-for-capm/stocks_dataset.csv')
# remove the date
df=df.iloc[:,1:]
df.head()
```

```python
returns=df.pct_change()
returns
```

## 2. Plot the prices and returns

```python
df.plot()
```

```python
# plot the 10 days FB , TWTR and NFLX returns 
days=10
num_stocks=3

returns.iloc[:days+1,:num_stocks].plot.bar()
```

```python
returns.std()
```

```python
returns.mean()
```

* it can be seen that FB and TWTR are more volatile than the NFLX during this period

## 3. How to compound the returns

```python
# numpy prod function

np.prod(returns+1)-1
```

```python
# alternative way : just use the built-in function: prod()
(returns+1).prod()-1
```

```python
# to round to 2 decimal, unit in %
compound=(returns+1).prod()-1
(compound*100).round(2)
```
