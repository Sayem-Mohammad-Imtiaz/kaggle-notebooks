# **Zomato Stock Price Prediction**

**We are given with data of stock prices and volumes of Zomato which was opened on 23rd July 2021 and the available dataset is upto 6th September 2021**

**Description of dataset**
1. Date : Date 
2. Open : Opening Price
3. Hight : Highest price in that day
4. Low : Lowest Price in that day
5. Close: Closing Price 
6. Volume : Volume available 

***Importing Libraries***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import math
import seaborn as sns
```

***Importing Dataset***

```python
zomato = pd.read_csv('../input/zomato-share-prices-jul-23-sept-07-2021/Zomato Share Prices till Date (Jul 23 - Sept 07 2021).csv')
```

```python
zomato
```

***Describing Dataset***

```python
zomato.info()
```

```python
zomato.describe()
```

```python
zomato.shape
```

**We will not need volume column so we are droping volume from dataset**

```python
zomato_df = zomato.drop('Volume', axis = 1)
zomato_df
```

**Droping Null Values**

```python
zomato_df1 = zomato.dropna(subset = ['High','Low','Close*','Open'])
```

**Variation in Opening prices**

```python
figg1 = plt.figure(figsize = (50,10))
sns.lineplot(x = 'Date', y ='Open', data = zomato)
```

**Variation in High Values**

```python
figg1 = plt.figure(figsize = (50,10))
sns.lineplot(x = 'Date', y ='High', data = zomato)
```

**Realation between open and High Prices**

```python
plt.scatter('Open','High', data = zomato)
plt.xlabel('Open Prices')
plt.ylabel('Highest ')
```

**Separating Data Set**

```python
x = zomato_df1[['High','Low','Open']].values
y = zomato_df1['Close*'].values
```

```python
x
```

```python
y
```

**Training and Testing Sets**

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
regressor = LinearRegression()
```

```python
regressor.fit(x_train, y_train)
```

```python
print(regressor.coef_)
```

```python
print(regressor.intercept_)
```

```python
predict = regressor.predict(x_test)
```

```python
print(predict)
```

```python
zomato1 = pd.DataFrame({'Actual_value': y_test.flatten(), 'Predicted_value' : predict.flatten()})
```

```python
zomato1.head(10)
```

```python
print('Mean Abs Error:', metrics.mean_absolute_error(y_test,predict))
print('Mean Sq Error:', metrics.mean_squared_error(y_test,predict))
print('Root Mean Sq Error:', math.sqrt(metrics.mean_squared_error(y_test,predict)))
```

```python
zomato1.plot(kind = 'line')
```

We can conclude that we are observing nearabout constant difference between actual values and predicted values.
