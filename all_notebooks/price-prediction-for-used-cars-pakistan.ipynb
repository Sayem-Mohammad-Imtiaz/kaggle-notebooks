# Importing necessary libraries for reading the csv

```python
import pandas as pd
import warnings
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
#Ignore the import and install warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('../input/pakistan-car-prices/updated_pakwheels.csv')
df.info()
```

### As the Car's make, model, year and variant is in the same column and is catagorical so we need to transform it into numerics

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
encoded_CarName = label_encoder.fit_transform(df.Name)
```

## Car names are now transformed

```python
encoded_CarName = pd.Series(encoded_CarName)
df['Name'] = encoded_CarName
```

## We have 7229 different cars available from 54913 ads

### We will now prepare our x and y variables for prediction where x is the list of independent variables and y is the dependent variable

```python
x = df[['Engine Capacity', 'Model Year', 'Mileage','Name', 'Engine Type', 'Transmission', 'Body Type']]
y = df['Price']
```

## Training the model on decision tree with the criterion as Mean Absolute Error

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

independent_train, independent_test, dependent_train, dependent_test = train_test_split(x, y, test_size = 0.30, random_state = 46)
regr = DecisionTreeRegressor(criterion='mae')
model = regr.fit(independent_train, dependent_train)
```

## As the model is trained now, we can check its accuracy below

```python
predicted_values = model.predict(independent_test)
r2_score(dependent_test, predicted_values)
```

## Here, we can check the first 5 real vs predicted values

```python
predict_df = pd.DataFrame({"Dependent_Test" : dependent_test, "Dependent_Predicted" : predicted_values})
predict_df.head()
```

## Training the model using KNN

```python
from sklearn import neighbors
model_knn = neighbors.KNeighborsRegressor(n_neighbors = 3)
model_knn.fit(independent_train, dependent_train)
```

## As the model is trained now wih KNN, we can check its accuracy below

```python
predicted_value = model_knn.predict(independent_test)
knn_score = r2_score(dependent_test, predicted_value)
knn_score
```

# R2 score with decision tree is better than KNN

```python
from sklearn.naive_bayes import GaussianNB
modelNB = GaussianNB()
modelNB.fit(independent_train, dependent_train)
```

## As the model is trained now with Naive bayes, we can check its accuracy below

```python
predicted_val = modelNB.predict(independent_test)

NB_score = r2_score(dependent_test, predicted_val)
NB_score
```
