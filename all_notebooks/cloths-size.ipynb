## 2. Load Data

```python
df = pd.read_csv('../input/clothessizeprediction/final_test.csv')
df.head() # Visualize first 5 rows of data
```

```python
df.describe()
```

## 3. Finding null values

```python
pd.isna(df).sum()
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
df['height'].fillna(df['height'].mean(), inplace=True)
```

```python
pd.isna(df).sum()
```

## 4. Visualize Data

```python
sns.countplot(df['size'])
plt.show()
```

## 5. Preprocessing of Data

```python
from sklearn.preprocessing import LabelEncoder
LBE = LabelEncoder()
df['size_class'] = LBE.fit_transform(df['size'])
```

```python
df.head()
```

```python
sns.countplot(df['size_class'])
plt.show()
```

```python
# Create target variable
Y = df['size_class']
Y.head()
```

```python
X = df.iloc[:,:3]
```

```python
standard = MinMaxScaler()
std_x = standard.fit_transform(X)
```

## 6. Split Data in Training & Testing samples

```python
X_train, X_test, y_train, y_test = train_test_split(std_x,Y,test_size=0.20,random_state = 40)
print('Shape of Training Xs:{}'.format(X_train.shape))
print('Shape of Test Xs:{}'.format(X_test.shape))
```

## 7. Build Logistic Regression model

```python
learner = LogisticRegression(max_iter=10000); #initializing Logistic Regression model

learner.fit(X_train,y_train); #training the Logistic Regression model
y_predicted = learner.predict(X_test)
score=learner.score(X_test,y_test);#testing the Logistic Regression model
```

## 8. Predict outcome

```python
print(score)
print(y_predicted)
```

```python
df_new = pd.DataFrame({"true_score":y_test,"predicted_score":y_predicted})
df_new
```
