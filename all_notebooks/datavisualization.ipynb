```python
train1=train[~train["Min_coverage"].isnull()]
print(train1.isnull().sum())
print("shape: ",train1.shape)
```

```python
train2=train1[~train1["Amount"].isnull()]
print(train2.isnull().sum())
print("shape: ",train2.shape)
```

```python
train2.corr("spearman")
```

```python
sns.countplot(x="Condition",data=train2)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Insurance_company",data=train2)
```

```python
crosstab=pd.crosstab(index=train2["Condition"],columns=train2["Insurance_company"])
pd.crosstab(index=train2["Condition"],columns=train2["Insurance_company"],margins=True)
```

```python
(chi2,p,dof,_)=stats.chi2_contingency([crosstab.iloc[0].values,crosstab.iloc[1].values])
```

```python
print("chi2: ",chi2)
print("p-value: ",p)
print("dof: ",dof)
```

```python
# chi2 > 5.99 for 5% level of significance, thus both are not independent of each other.
#strength of these variables (correlation)
np.sqrt(chi2/train2.shape[0])
```
