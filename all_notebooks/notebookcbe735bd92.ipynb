importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
data = pd.read_csv('../input/headbrain-linear-regression/headbrain_dataset.csv')
print(data.shape)
data.head()
```

assigning values

```python
X= data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values
```

finding mean and value of c and m

```python
mean_X = np.mean(X)
mean_Y = np.mean(Y)

n=len(X)
numer= 0
denom =0
for i in range(n):
    numer += (X[i] - mean_X)* (Y[i]-mean_Y)
    denom += (X[i] - mean_X) ** 2
b1 = numer/denom
b0 = mean_Y - (b1 * mean_X)
print(b1,b0)
```

plotting 

```python
max_X=np.max(X) + 100
min_X=np.min(X) - 100
X=np.linspace(min_X, max_X, 1000)
Y = b0 + b1 * X
plt.plot(X,Y, color='#58b970', label='Regression Line')
plt.scatter(X,Y, color='#ef5423', label='scatter plot')
plt.xlabel('Head Size cm^3)')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```
