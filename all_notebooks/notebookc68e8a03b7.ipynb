```python
df=pd. read_csv("../input/up-crime-data-year-2019/MR data - Compiled Data Set.csv")
df=df.drop(df.columns[[1]], axis=1)
df=df.drop(df.columns[[1]], axis=1)
df.head()
```

```python
newcols = {
    'Number of male(adult)': 'Km', 
    'Number of adult victims': 'Ca', 
    'Number of childs invoved': 'Cch',
    'Number of child victims': 'Mch', 
    'Number of male victims(adult)': 'Mm', 
    'Kidnnaping:\nNumber of child victims': 'Kch',
    'Number of male(adult) ': 'Km', 
    'Number of female victims(adult)': 'Mf', 
    'Number of female(adult)': 'Kf',
    'Murder Happened with reason:\n1:Property/Land Disputes,\n2:Family Dispute\n3:Petty Quarrels,\n4:Money Disputes, \n5:Personal Vendetta, \n6:Love Affairs, \n7:Casteism\n8: Unknown/other': 'M', 
    'Crime Against Women ( Combined ):\n1. Murder with Rape \n2. Dowry Deaths(Sec. 3048)\n3. Suicide(sec 305/306)\n4. Kidnapping(All)\n5. Acid Attack(Sec. 326A IPC)\n6. Cruelty by Husband/in-laws((Sec.498 A IPC)\n7. Rape only(Sec. 376 or 511 IPC)\n8. Assault on Women with Intent to Outrage her Modesty (Sec. 354 IPC)\n9. Cyber Crimes against Women\n10. Protection of Children from Sexual Offences Act': 'C'    
}
df.rename(columns=newcols, inplace=True)
df.head()
```

```python
df[["l","m","n"]]=df["C"].str.split(",", n = 2, expand = True)
df=df.replace(np.nan, 0)
df.head()
```

```python
df['Mch'] = df['Mch'].apply(int)
df['Mm'] = df['Mm'].apply(int)
df['Mf'] = df['Mf'].apply(int)
df['Kch'] = df['Kch'].apply(int)
df['Cch'] = df['Cch'].apply(int)
df['Km'] = df['Km'].apply(int)
df['Kf'] = df['Kf'].apply(int)
df['Ca'] = df['Ca'].apply(int)
df['l'] = df['l'].apply(int)
df['m'] = df['m'].apply(int)
df['n'] = df['n'].apply(int)
print(df)
```

```python
for i in range (0,len(df)):
    if ',' in str(df.iloc[i,8]):
        df.iloc[i,8]=0;
df['C'] = df['C'].apply(int)   
df['l']=df['l']- df['C'];
print(df)
```

```python
d={0:0,'Property Disputes':1,'Family Dispute':2,'Petty Quarrels':3,'Money Disputes':4,'Personal Vendetta':5,'Love Affairs':6,'Casteism':7,'Unknown reasons':8}
df['M']=df['M'].apply(lambda x : d[x])
```

```python
GM1 = df['Mch'][df.iloc[:,0]=='Ghaziabad'].sum()
GK1 = df['Kch'][df.iloc[:,0]=='Ghaziabad'].sum()
GC1 = df['Ca'][df.iloc[:,0]=='Ghaziabad'].sum()
KM1 = df['Mch'][df.iloc[:,0]=='Kanpur'].sum()
KK1 = df['Kch'][df.iloc[:,0]=='Kanpur'].sum()
KC1 = df['Ca'][df.iloc[:,0]=='Kanpur'].sum()
LM1 = df['Mch'][df.iloc[:,0]=='Lucknow'].sum()
LK1 = df['Kch'][df.iloc[:,0]=='Lucknow'].sum()
LC1 = df['Ca'][df.iloc[:,0]=='Lucknow'].sum()
GM2 = df['Mch'][df.iloc[:,0]=='Ghaziabad'].sum()
GK2 = df['Kch'][df.iloc[:,0]=='Ghaziabad'].sum()
GC2 = df['Ca'][df.iloc[:,0]=='Ghaziabad'].sum()
KM2 = df['Mch'][df.iloc[:,0]=='Kanpur'].sum()
KK2 = df['Kch'][df.iloc[:,0]=='Kanpur'].sum()
KC2 = df['Ca'][df.iloc[:,0]=='Kanpur'].sum()
LM2 = df['Mch'][df.iloc[:,0]=='Lucknow'].sum()
LK2 = df['Kch'][df.iloc[:,0]=='Lucknow'].sum()
LC2 = df['Ca'][df.iloc[:,0]=='Lucknow'].sum()
GM3 = df['Mch'][df.iloc[:,0]=='Ghaziabad'].sum()
GK3 = df['Kch'][df.iloc[:,0]=='Ghaziabad'].sum()
KM3 = df['Mch'][df.iloc[:,0]=='Kanpur'].sum()
KK3 = df['Kch'][df.iloc[:,0]=='Kanpur'].sum()
LM3 = df['Mch'][df.iloc[:,0]=='Lucknow'].sum()
LK3 = df['Kch'][df.iloc[:,0]=='Lucknow'].sum()
GM=GM1+GM2+GM3
GK=GK1+GK2+GK3
KK=KK1+KK2+KK3
GC=GC1+GC2
KM=KM1+KM2+KM3
KC=KC1+KC2
LM=LM1+LM2+LM3
LK=LK1+LK2+LK3
LC=LC1+LC2

Mcr = [GM,KM,LM]
Kcr= [GK,KK,LK]
Ccr= [GC,KC,LC]
```
