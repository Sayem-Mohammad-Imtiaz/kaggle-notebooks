So what are the top **PROTEIN** sources (Category)?

```python
pd.pivot_table(df,index=['Category'],values=['Protein'],aggfunc=np.max).plot(kind='bar')
```

We want to get a lot of Protein but we do not want the sugar! This would ruin our amazing beach body-shape! So Protein MAX and Sugar MIN! What are our best options?

```python
df['Protein/Sugar'] = np.where(df['Sugars'] < 1, df['Sugars'], df['Protein']/df['Sugars'])
```

```python
df.sort_values(by='Protein/Sugar', ascending=False).head(10)
```

Okay, Chicken McNuggets 40 piece is still the Top-Dog! But on for the silver and bronze medal the Breakfast-squad is representing! Sausage McMuffin with Egg and Sausage McMuffin with Egg Whites are giving you a great bang for your buck!
