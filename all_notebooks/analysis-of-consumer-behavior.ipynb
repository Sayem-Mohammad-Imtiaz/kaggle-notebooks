Viendo las columnas del DataFrame.

```python
Data=pd.read_csv("../input/Mall_Customers.csv")
Data.head()
```

```python
Data.describe()
```

Descripción general de los datos.

```python
plt.hist(Data['Age'])
plt.xlabel('Age')
plt.show()
plt.hist(Data['Annual Income (k$)'])
plt.xlabel('Annual Income (k$)')
plt.show()
plt.hist(Data['Spending Score (1-100)'])
plt.xlabel('Spending Score (1-100)')
plt.show()
```

Mayor parte de la gente que hace compras se encuentra entre el rango de edad de 20 a 50 años. La gran mayoria de personas cuenta con un salario anual por debajo de los $80,000 y el spending score se encuentra muy equilibrado con una predominancia en el rango de 40 a 60 puntos. 
Es necesario seguir indagando en los datos.

```python
Muestra=Data['Gender'].value_counts()
print('%Female:',Muestra[0]*100/(Muestra[0]+Muestra[1]))
print('%Male:',Muestra[1]*100/(Muestra[0]+Muestra[1]))
```

Mayor porcentaje de las compras las efectuan las mujeres. Usualmente las mujeres cumplen el rol de cuidado primario en su circulo familiar, por lo cual generalmente hacen las compras para la sustentación de su familia. Mas mujeres visitan el Mall que los hombres.

```python
plt.figure(figsize=(15,5))
sns.boxplot(Data['Gender'], Data['Spending Score (1-100)'])
plt.title('Gender vs Spending Score')
plt.show()
```

Las mujeres aportan mas al 'Spending Score' que los hombres. Las mujeres gastan mas en promedio.

```python
plt.figure(figsize=(15,5))
sns.boxplot(Data['Gender'], Data['Annual Income (k$)'])
plt.title('Gender vs Annual Income')
plt.show()
```

Sin embargo, los clientes masculinos tienen más ingresos anuales en comparación con los clientes femeninos.
INTERESANTE.

```python
plt.scatter(Data['Annual Income (k$)'], Data['Age'],  c='red')
plt.ylabel('Age')
plt.xlabel('Annual Income (k$)')
plt.show()
plt.scatter(Data['Spending Score (1-100)'], Data['Age'],  c='red')
plt.ylabel('Age')
plt.xlabel('Spending Score (1-100)')
plt.show()
plt.scatter(Data['Spending Score (1-100)'], Data['Annual Income (k$)'],  c='red')
plt.ylabel('Annual Income (k$)')
plt.xlabel('Spending Score (1-100)')
plt.show()
```

1) A medida que disminuye la edad del individuo el ingreso anual va aumentando en promedio, pero luego en el rango de edad 18-25 vuelve a bajar el inreso anual.

2) La gente por debajo de los 40 años tiene un mayor 'spending score' en promedio que el otro grupo de edad. Es decir, la población mas joven, gasta mas.

3) Se evidencia una segmentación de clusters poblacionales que tienen distintos comportamientos de consumo relacionados a su ingreso anual. Entonces, se puede decir que hay 5 grupos objetivo basados en su ingreso y gasto, mas no en su edad:

- Bajo ingreso / Bajo gasto
- Bajo ingreso / Alto gasto
- Ingreso medio / Gasto medio
- Alto ingreso / Bajo gasto
- Alto ingreso / Alto gasto

Basados en estos 5 grupos se pueden hacer estrategias de marketing focalizadas a cada grupo y asi mejorar el 'spending score' o tambien para mantener el consumo, como programas de fidelización o cupones de descuento. Tambien se pueden plantear nuevas preguntas de investigación como: ¿Por que hay gente con un alto ingreso pero consume poco?  ¿será un nicho de edad específico? 


```python
Corr1=Data['Spending Score (1-100)'].corr(Data['Annual Income (k$)'])
print('CORRELACION gasto-ingreso :', Corr1*100,'%')
Corr2=Data['Spending Score (1-100)'].corr(Data['Age'])
print('CORRELACION gasto-edad :', Corr2*100,'%')
Corr3=Data['Annual Income (k$)'].corr(Data['Age'])
print('CORRELACION ingreso-edad :', Corr3*100,'%')
```

Finalmente, no hay una relación fuerte entre la edad y el ingreso y este ultimo realmente no explica el comportamiento de consumo, y esto lo vimos anteriormente con los distintos clusters de consumidores donde no se ve un patrón donde a mayor ingreso mayor gasto.
Por otro lado, se evidencia una relación mas significativa negativamente entre gasto y edad lo cual refuerza lo dicho antes. Esto sirve para poner especial atención a la población de mayor edad y buscar formas de atraer su consumo e igualmente para la población joven de altos ingresos pero que tiene un bajo consumo.
