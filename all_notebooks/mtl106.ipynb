```python
def plot(df):
    df=pd.DataFrame(df)
    for i in range(len(df.columns)):
        col=df.iloc[:,i]
        if(col.value_counts().count()>40 and (isinstance(col[0], np.float64)==True or isinstance(col[0],np.int64)==True)):
          col=col.dropna()
          if(sum(col==0)>0.75*len(col)):
             sns.distplot(col,kde_kws={'bw':1});
             plt.show()
          else:
             sns.distplot(col);
             plt.show()
    return 

```

```python
d=df.groupby(['month','date_of_month'],as_index=False).sum()
d
```

```python
d['births']=d['births']/d['births'].sum()
```

```python
#d.drop(['month','date_of_birth'],axis=1,inplace=True)
pro=d['births']
```

```python
plt.plot(pro)
```

```python
pro=np.cumsum(pro)
```

```python
pro
```

```python
from bisect import bisect_left 
import random
def bs(a, x): 
    i = bisect_left(a, x) 
    if i: 
        return (i-1) 
    else: 
        return -1
s=set()
times = 10000
count=0
for i in range(times):
    for j in range(23):
        val=bs(pro,random.random())
        #print(val)
        if(val in s):
            count+=1
            s.clear()
            break
        s.add(val)
count
```
