```python
TenneTTSO=pd.read_csv('/kaggle/input/wind-power-generation/TenneTTSO.csv')
Hertz=pd.read_csv('/kaggle/input/wind-power-generation/50Hertz.csv')
TransnetBW=pd.read_csv('/kaggle/input/wind-power-generation/TransnetBW.csv')
Amprion=pd.read_csv('/kaggle/input/wind-power-generation/Amprion.csv')
```

How time measurements correlate with each other? Significant bias in TransnetBW is visible. 

Amprion

```python
#Consider how time correlated
Amprion.set_index('Date',inplace=True)
import plotly.graph_objects as go
x=list(Amprion.corr().index)
y=list(Amprion.corr().columns)
values=np.array(Amprion.corr().values)
fig = go.Figure(data=go.Heatmap(
    z=values,
    x=x,
    y=y,
                 
    
                   hoverongaps = False))
fig.show()
```

```python
TransnetBW.set_index('Date',inplace=True)
import plotly.graph_objects as go
x=list(TransnetBW.corr().index)
y=list(TransnetBW.corr().columns)
values=np.array(TransnetBW.corr().values)
fig = go.Figure(data=go.Heatmap(
    z=values,
    x=x,
    y=y,
                 
    
                   hoverongaps = False))
fig.show()
```

So, wind changes by time so do productivity. However it's hard to say what time is the best but it's definitely not early in the morning. 

```python
import matplotlib.pyplot as plt
std=pd.DataFrame({'TenneTTSO':np.std(TenneTTSO),'Hertz':np.std(Hertz),'TransnetBW':np.std(TransnetBW),'Amprion':np.std(Amprion)})
summ=pd.DataFrame({'TenneTTSO':TenneTTSO.median(),'Hertz':Hertz.median(),'TransnetBW':TransnetBW.median(),'Amprion':Amprion.median()})
summ.plot(figsize=(10,10),yerr=std)
```
