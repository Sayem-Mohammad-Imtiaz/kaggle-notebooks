# Ham or Spam classification
The data set has two columns one is the subject of the mail and the other is the label called spam, which specifies if the given document is ham or spam. The value 1 indicates it's a spam and vice versa.

This is typically a classification problem and since it has only labels either 0 or 1 it falls under the binary classification

# train & test data split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.text,df.spam,test_size=0.2,random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

# Tokenization & Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
vectorizer.fit_transform(X_train)
print(f"the total vocabulary size in the train data is {len(vectorizer.vocabulary_)}")
```

```python
vectorizer.vocabulary_
#vocabulary:index
```

```python
vectorizer.get_feature_names()
```

```python
X_train_ct = vectorizer.transform(X_train)
```

```python
X_train_ct.shape
```

```python
X_train_ct[0]
```

```python
print(X_train_ct[0])
```

```python
X_train[0]
```

```python
X_test_ct =  vectorizer.transform(X_test)
```

```python
print(X_test_ct[0].shape)
```

```python
print(X_test_ct[0])
```

# Knn classifier

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train_ct,y_train)
```

```python
y_pred = knn_classifier.predict(X_test_ct)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

# SVC 

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train_ct,y_train)
y_pred = svc.predict(X_test_ct)
accuracy_score(y_test,y_pred)
```

# Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
nb= MultinomialNB()
nb.fit(X_train_ct,y_train)
y_pred = nb.predict(X_test_ct)
accuracy_score(y_test,y_pred)
```

# TF-IDF Vectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
vectorizer.fit(X_train)
X_train_ct  = vectorizer.transform(X_train)
X_test_ct = vectorizer.transform(X_test)
```

# KNN

```python
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train_ct,y_train)
y_pred = knn_classifier.predict(X_test_ct)
accuracy_score(y_test,y_pred)
```

# SVC

```python
svc = SVC()
svc.fit(X_train_ct,y_train)
y_pred = svc.predict(X_test_ct)
accuracy_score(y_test,y_pred)
```

# Naive Bayes

```python
nb= MultinomialNB()
nb.fit(X_train_ct,y_train)
y_pred = nb.predict(X_test_ct)
accuracy_score(y_test,y_pred)
```

# TF-IDF with n-gram

```python
vectorizer = TfidfVectorizer(ngram_range=(1,2))
vectorizer.fit(X_train)
X_train_ct = vectorizer.transform(X_train)
X_test_ct = vectorizer.transform(X_test)
```

# KNN

```python
knn_classifier = KNeighborsClassifier()
knn_classifier.fit(X_train_ct,y_train)
y_pred = knn_classifier.predict(X_test_ct)
accuracy_score(y_test,y_pred)
```

# SVC

```python
svc = SVC()
svc.fit(X_train_ct,y_train)
y_pred = svc.predict(X_test_ct)
accuracy_score(y_test,y_pred)
```

# Naive Bayes

```python
nb= MultinomialNB()
nb.fit(X_train_ct,y_train)
y_pred = nb.predict(X_test_ct)
accuracy_score(y_test,y_pred)
```

```python
import tensorflow as tf
tf.keras.backend.clear_session
```

```python
model = tf.keras.Sequential()
model.add(tf.keras.layers.Dense(100,activation='relu',input_shape=(len(vectorizer.vocabulary_),)))
model.add(tf.keras.layers.Dense(50,activation='relu'))
model.add(tf.keras.layers.Dropout(0.4))          
model.add(tf.keras.layers.Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['acc'])          
```

```python
model.fit(X_train_ct.todense(),y_train,validation_data=(X_test_ct.todense(),y_test),epochs=3,batch_size=32)
```
