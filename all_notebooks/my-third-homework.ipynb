**.melt() tidies unclean datas.**

```python
data_new = data.head()
data_new
```

**frame = source data frame, id_vars= no melted feature, value_vars= melted features**

```python
melted = pd.melt(frame = data_new, id_vars = 'Country', value_vars = ['Population', 'GDP ($ per capita)','Birthrate'])
melted
```

**Pivoting is reverse of the melting.**
* Index is assigned what we want to replace the index.
* columns like a keys in dictionary.
* values like a values in dictionary.

```python
melted.pivot(index = 'Country', columns = 'variable', values = 'value')
```

**Concatenating the two datas with pd.concat() method.**

```python
data1 = data.head()
data2 = data.tail()

data_conc_row = pd.concat([data1,data2], axis = 0, ignore_index = True) # axis = 0 concatenates dataframes in row.
data_conc_row
```

```python
data1 = data['Population'].head()
data2 = data['Birthrate'].head()

data_conc_column = pd.concat([data1,data2], axis = 1) # axis = 1 concatenates dataframes in column.
data_conc_column 
```
