```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (32, 14)
plt.rcParams["font.size"] = 23

f1 = open("/kaggle/input/abortion-statistics-year-ended-december-2019/as-dec-19-general-abortion-rate.csv", "r")
f2 = open("/kaggle/input/abortion-statistics-year-ended-december-2019/as-dec19-abortion-rates-by-age-of-woman.csv", "r")

lines1 = f1.readlines()
lines2 = f2.readlines()

print(lines2[0])

years = ['1980', '1981', '1982', '1983', '1984', '1985', '1986', '1987', '1988', '1989', '1990', '1991', '1992', '1993', '1994', '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003', '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']
outs = [0]*len(years)

for i in range(1, len(lines1)):
    line = lines1[i].replace("\n", "").split(",")
    outs[years.index(line[0])] += float(line[1])

plt.xticks(rotation=75)
plt.title("General abortion rate by years")
plt.xlabel("Years")
plt.ylabel("General abortion rate")
plt.grid(linestyle="--")
plt.plot(years, outs, marker="o", linewidth=1.718, color="orange")
plt.show()

ages = [' 11-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40-44', '45 and over']
outs = [0.0]*len(ages)
years = [0.0]*len(ages)

for i in range(1, len(lines1)):
    line = lines2[i].replace("\n", "").split(",")
    
    outs[ages.index(line[1])] += float(line[2])
    years[ages.index(line[1])] += (int(line[0])*282)/2019

plt.title("Abortion rates by age of woman")
plt.xlabel("Ages")
plt.ylabel("Abortion rate")
plt.grid(linestyle="--", linewidth=1.618)
plt.bar(ages, outs, linewidth=1.618)
plt.show()
    
```
