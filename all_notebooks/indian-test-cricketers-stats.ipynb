In the dataset to calculate batsman's average more accurately, I had to clean the data["Mat"] which had "[" in between the string at few spots, for which I used this function.

```python
import string
print(string.punctuation)
```

```python
def clear(num):
    numx = list(num)
    coun = len(numx)
    ao = []
    x = ""
    
    for i in range(coun):
        if numx[i] in string.punctuation:
            for j in range(0,i):
                ao.append(numx[j])
                
            for i in ao:
                x = x + i
            
            return x

    return num
```

```python
n = []
for i in data_tp["Mat"]:
    n.append(clear(i))
    
    
data_tp["Matches"]  = pd.DataFrame(n)

data_tp["Matches"] = data_tp["Matches"].astype(float)


    
```

```python
data_tp[["Name","Matches"]]
```

```python
batsman_rating = round((data_tp["Runs"])/(data_tp["Matches"]) + data_tp["Matches"]*0.25)
data_tp["Bat_rating"] = batsman_rating
```

```python
data_bat = data_tp[["Name","Bat_rating"]].sort_values(by = ["Bat_rating"],ascending = False)[0:10]
```

```python
plt.figure(figsize = (10,7))
sns.barplot(y = "Name",x = "Bat_rating",data = data_bat)
plt.title("Batsman Performance Rating (Top 10)")
```

```python
data_bowl = data_tp[["Name","Ave"]].replace("-","0",regex = True)
data_bowl["Ave"] = data_bowl.Ave.astype(float)
data_bowl = data_bowl.sort_values(by = ["Ave"])
data_bowl
```

```python
data_bowl["Ave"] = data_bowl["Ave"].replace(0,(data_bowl["Ave"].mean()+100))
```

```python
data_bowl["Matches"] = data_tp["Matches"]
```

```python
data_bowl.sort_values(by = ["Ave"])
```

```python
data_bowl["Wicket"] = data_tp["Wkt"]
```

```python
bowl_rating = (data_bowl["Matches"]*0.25 + data_bowl["Wicket"]*1)/data_bowl["Ave"]
data_bowl["Bowl_rating"] = bowl_rating
```

```python
data_bowl = data_bowl.sort_values(by = ["Bowl_rating"],ascending = False)
data_bowl =data_bowl[:20]

```

```python
plt.figure(figsize = (10,7))
sns.barplot(x = "Name",y = "Bowl_rating",data = data_bowl,palette="Blues_d")
plt.xticks(rotation = 90)
plt.title("Top 20 Indian Test Cricket Bowlers")
```
