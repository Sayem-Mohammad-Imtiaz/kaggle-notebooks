# 1. Import neccessary Library.

```python
#to store and process data in dataframe
import pandas as pd

#for numeriacal analysis
import numpy as np

#for advance visualization
import seaborn as sns

# for basic visualization
import matplotlib.pyplot as plt
```

# 2. Dataset

```python
df = pd.read_csv('/kaggle/input/covid19-by-country-with-government-response/covid19_by_country.csv')

df['Date'] = pd.to_datetime(df['Date'], format = '%Y-%m-%d')
```

# 3. Filter Country

only ASEAN countries:  10 members

![](https://sites.google.com/site/saowakonsuedee/_/rsrc/1391927199702/about-10-asean-countries-history-of-the-10th-asean-countries-1/asean1.jpg?height=370&width=400)

Source: https://sites.google.com/site/saowakonsuedee/about-10-asean-countries-history-of-the-10th-asean-countries-1


```python
df_asean = df[(df['Country']=='Thailand')|
              (df['Country']=='Singapore')| 
              (df['Country']=='Laos')|
              (df['Country']=='Myanmar')|   
              (df['Country']=='Vietnam')|
              (df['Country']=='Malaysia')|
              (df['Country']=='Brunei')|  
              (df['Country']=='Indonesia')|     
              (df['Country']=='Philippines')|
              (df['Country']=='Cambodia')              
              ]  
# check Countries
df_asean['Country'].value_counts()
```


**Right now, there is no Myanmar,**

I think, The current situation of unrest in the country.

# 4. Filter Time

I need to review only FY21

```python
df_asean2021 = df_asean[df_asean['Date'] > '2020-12-31']
```

# 5 Review

```python
# Total confirm case

fig, ax = plt.subplots( 1,1,figsize=(15,6))
sns.lineplot(x='Date', y='confirmed', data = df_asean2021 , hue='Country')
plt.title('Total cases')
plt.ylabel(' Counts x 1,000,000')
```

```python
# Total confirm Deaths

fig, ax = plt.subplots( 1,1,figsize=(15,6))
sns.lineplot(x='Date', y='deaths', data = df_asean2021 , hue='Country')
plt.title('Total Deaths')
plt.ylabel(' Counts ')
```

```python
# Daily confirm case

fig, ax = plt.subplots( 1,1,figsize=(15,6))
sns.lineplot(x='Date', y='confirmed_inc', data = df_asean2021 , hue='Country')
plt.title('Daily New Cases')
plt.ylabel(' Counts')
```

```python
# Daily confirm Deaths

fig, ax = plt.subplots( 1,1,figsize=(15,6))
sns.lineplot(x='Date', y='deaths_inc', data = df_asean2021 , hue='Country')
plt.title('Daily New Deaths')
plt.ylabel(' Counts')
```
