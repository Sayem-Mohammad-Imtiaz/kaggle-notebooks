That does not look, good, maybe throw out these notes, like Verse, Chorus..., just everything that is in these square brackets. With this regex: \[.*?\]

```python
import re
clean_all_lyrics = []
for lyric in all_lyrics:
    entry = re.sub("\[.*?\]","",lyric)
    clean_all_lyrics.append(entry)
```

Looks good, lets try the next word cloud on it.

```python
# get html code
embed_code=wc.get_embed_code(text=clean_all_lyrics,random_color=True,topn=24)

# display
HTML(embed_code)
```

That looks nice, but I wanna do only a percentage of words.

```python
total_word_count = 0
for clean_lyric in clean_all_lyrics:
    split_string = clean_lyric.split()
    #print(split_string)
    word_count = len(split_string)
    #print(word_count)
    total_word_count += word_count
print(total_word_count)
word_cloud_topn = int(np.floor(total_word_count * 0.001))
print(word_cloud_topn)
```

```python
# get html code
embed_code=wc.get_embed_code(text=clean_all_lyrics,random_color=True,topn=word_cloud_topn)

# display
HTML(embed_code)
```
