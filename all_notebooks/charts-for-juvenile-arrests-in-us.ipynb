```python
data = pd.read_csv("../input/us-juvenile-arrests-by-crime/arrests_national_juvenile.csv", parse_dates = True, index_col = "year")
data.head()
```

```python
data.state_abbr
data = data.drop('state_abbr', axis=1)
```

```python
print(data.info())
print('\n')
cols_with_missing = []
object_cols = []

for col in data.columns:
    if data[col].isnull().any():
        cols_with_missing.append(col)
print("columns with missing values - "+ str(cols_with_missing)) #columns with missing values
print('\n')
for col in data.columns:
    if data[col].dtype == 'object':
        object_cols.append(col)
print("columns with categorical variables - "+str(object_cols))#columns with categorical variables
print('\n')
numeric_cols = set(data.columns) - set(object_cols) - set(data['id'])
print("columns without categorical variables - "+str(numeric_cols))#columns without categorical variables
print('\n')
print("unique values for offense - "+str(data['offense_name'].unique()))#unique values for offense code column (unique crimes)
print('\n')
print("number of such unique values - "+str(data['offense_name'].nunique()))#number of such unique values
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data_without_cat = data.drop(object_cols, axis=1) # data without categorical variables
data_without_cat
data_cat = data.drop(numeric_cols, axis=1) # data with only categorical variables
data_cat['id'] = data.id
data_cat
```

```python
plt.figure(figsize = (15,15))
sns.lineplot(data = data_without_cat['total_female'], label = "females arrested")
sns.lineplot(data = data_without_cat['total_male'], label = "males arrested")
plt.legend()
```
