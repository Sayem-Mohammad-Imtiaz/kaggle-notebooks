```python
import pandas as pd
df = pd.read_csv('../input/datadiamonds/datadiamonds.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
df.head()
```

**Process Grouping**

Jelaskan prosess yang terjadi dalam grouping data menggungkan method groupby() dari pandas.

1. Splitting : Memisahkan data kedalam suatu group berdasarkan kriteria tertentu.
2. Applying : Melakukan suatu operasi terhadap sekumpulan data di group-group tersebut.
3. Combining : Menggabungkan data menjadi suatu struktur baru

**Grouping All Column**

Group data diamond berdasarkan cut (kualitas), lakukan aggregasi dengan fungsi rata2.
Kemudian jawab pertanyaan berikut.

Berapa harga rata2 dari diamond berkualitas Premium?

```python
df.groupby('cut').mean()
```

```python
df.groupby('cut').mean().loc['Premium']
```

**Grouping Some Column base on Multiple Criteria**

Lakukan groupby pada data berdasarkan kriteria cut(kualitas) dan color (warna) secara berurutan, dengan aggregasi pada column carat, table, dan price saja. dengan fungsi agregasi rata-rata.Kemudian jawab pertanyaan berikut :

1. Berapa harga diamond yang berkriteria premium dan berwarna J
2. Berapa nilai dari carat diamond yang berkriteria very good dan berwarna H

```python
df.groupby(['cut','color'])[['carat','table','price']].mean()
```

```python
#harga diamond yang berkriteria premium dan berwarna J
df.groupby(['cut','color'])[['price']].mean().loc['Premium','J']
```

```python
#nilai dari carat diamond yang berkriteria very good dan berwarna H
df.groupby(['cut','color'])[['carat']].mean().loc['Very Good','H']
```

**Group By Multiple Aggregate Base on Multiple Criteria**

Lakukan groupby pada data berdasarkan kriteria cut(kualitas) dan color(warna) secara berurutan kemudian aggregasi objek group by nya dengan fungsi rata2 dan median.Kemudian Jawab Pertanyaan berikut :

Berapa nilai median dari harga diamond berkualitas good dan berwarna F

```python
# code here
df.groupby(['cut','color']).agg(['mean','median'])
```

```python
#nilai median dari harga diamond berkualitas good dan berwarna F
df.groupby(['cut','color'])[['table']].median().loc['Good','F']
```

**Group By With Different Aggregation for specific Column**

Lakukan groupby pada data berdasarkan criteria cut(kualitas), kemudian aggreagate dengan fungsi max pada column price dan fungsi min pada column carat. Kemudian Jawab Pertanyaan berikut :

Berapa nilai max dari price untuk diamond berkualitas Ideal

```python
df.groupby('cut')[['price','carat']].agg({'price':'max','carat':'min'})
```

```python
#nilai max dari price untuk diamond berkualitas Ideal
df.groupby('cut')[['price']].max().loc['Ideal']
```

**Group By With Custom Function**

Lakukan groupby pada data berdasarkan criteria cut(kualitas), kemudian lakukan aggregasi dengan fungsi standar deviasi pada column carat dan fungsi rentang nilai antara max dan min pada column price. Kemudian jawab pertanyaan berikut :

Berapa selisih harga tertinggi dan terendah dari diamond berkualitas Premium

```python
def diff_max_min(x):
  return x.max() - x.min()

df.groupby('cut')[['price','carat']].agg({'price':diff_max_min,'carat':'std'})
```
