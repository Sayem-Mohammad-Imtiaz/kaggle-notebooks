```python
df = pd.read_csv("../input/weather-dataset-rattle-package/weatherAUS.csv")
df
```

# Data Exploring

```python
df = df.drop(["Evaporation","Sunshine"],axis= 1)
df
```

```python
k = 0
for l in df.isnull().sum().tolist():
    k += 1
    if l > 0:
        df = df.dropna(subset = [df.columns[k]])

df
```

```python
df.dropna(subset = ["MinTemp"],inplace = True)
df
```

## *Checking if there is NAN value remaining*

```python
df.isnull().sum()
```

## **Substracting the non-raining days**

```python
df_raining_days = df[df["RainToday"] == "Yes"]
df_raining_days
```

```python
df_raining_days = df_raining_days.drop(["WindGustDir","WindDir9am","WindDir3pm"],axis = 1)
df_raining_days
```

```python
df.Location.unique()
```

## *Now pick some Locations to work on*

```python
Albury = df_raining_days[df_raining_days["Location"] == "Albury"]
Albury
Albury.sort_values(by='Rainfall', ascending=False, na_position='first')
```

```python
sns.lineplot(x = "WindGustSpeed",y = "Rainfall", data = Albury);
```

```python
sns.lineplot(x = "Humidity9am",y = "Rainfall", data = Albury);
```

```python
sns.lineplot(x = "Cloud9am",y = "Rainfall", data = Albury);
```
