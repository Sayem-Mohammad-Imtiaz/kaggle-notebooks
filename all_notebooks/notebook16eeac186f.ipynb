```python
state_level_latest.fillna('No Notes', inplace=True)

state_level_latest.describe()
```

```python
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,10)
y = np.array(range(len(state_level_latest.State)))
x= state_level_latest['Confirmed']
plt.yticks(y, state_level_latest['State'])
plt.xlabel('Cases')
plt.plot( x, y,label = 'Confirmed Cases')
plt.plot(state_level_latest['Recovered'], y, label='Recovered')
plt.ylabel('States')
plt.title('State Level Latest')
plt.legend()
plt.grid()
plt.show()
# print(state_level_latest['Confirmed'])
# arr = np.array(range(len(state_level_latest['State'])))
# d = np.array(range(len(state_level_latest.State)))
# d
```

```python
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,10)
x = np.array(range(len(state_level_latest.State)))
y= state_level_latest['Confirmed']
plt.xticks( x, state_level_latest['State_code'])
plt.xlabel('States')
plt.bar(x, y, label='Confirmed Cases')
plt.bar(x, state_level_latest['Recovered'], label = 'Recovered')
plt.bar(x, state_level_latest['Active'], label = 'Active')
plt.bar(x, state_level_latest['Deaths'], label = 'Deaths')
plt.bar(x, state_level_latest['Migrated_Other'], label = 'Migrated')
plt.ylabel('Cases')
plt.title('State Level Latest')
plt.legend()
plt.show()
```
