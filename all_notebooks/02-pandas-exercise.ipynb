#### 4- آیا داده گمشده (missing data) وجود دارد؟ اگر جواب مثبت است، کدام ستون بیشترین تعداد داده گمشده را دارد؟

```python
# way 1
missing_datas = df.isnull().sum()

print(missing_datas)
```

```python
# way 2
max_value = missing_datas.max()
max_value_title = missing_datas.idxmax()

print(f'column "{max_value_title}" with ({max_value}) have most missing values')
```

```python
# way 3
max_value = -1
max_value_title = '-1'

for title in missing_datas.index:
    value = missing_datas[title]
    if max_value < value:
        max_value = value
        max_value_title = title
        
print(f'column "{max_value_title}" with ({max_value}) have most missing values')
```

```python
# way 4
missing_datas_info = dict()

for column in df.columns:
    missing_data_count = 0
    
    for value in df[column]:
        
        if value is np.NaN :
            missing_data_count+=1
    
    missing_datas_info[column] = missing_data_count
    
print(missing_datas_info)
print(max(missing_datas_info.items(),key = lambda item:item[1]))
```

#### 7- نام فردی که بیشترین پرداختی )ADR(Average Daily Rate را دارد مشخص کنید. مبلغ آن چقدر است؟

```python
# way 1
colume = 'adr'
max_value = df[colume].max()
index_max_value = df[colume].idxmax()
index_name = df['name'][index_max_value]

print(f'"{index_name}" with index({index_max_value}) have maximum value "{max_value}" in colume "{colume}"')
```

#### 13- کد تلفن مناطقی که بیشترین رزرو را داشته اند، مشخص کنید (3 کد پرتکرار)

```python
# way 1
df['phone-code'] = df['phone-number'].apply(lambda phone:phone.split('-')[0])

df.groupby('phone-code').size().sort_values(ascending=False)
```

```python
# way 2
phone_repeats = dict()

for phone in df['phone-number']:
    phone_code = phone.split('-')[0]
    
    if phone_code not in phone_repeats:
        phone_repeats[phone_code] = 0
        
    phone_repeats[phone_code] +=1
    
phone_repeats = sorted(phone_repeats.items(), key=lambda phone_repeat: phone_repeat[1])    
print(phone_repeats[:3])
```
