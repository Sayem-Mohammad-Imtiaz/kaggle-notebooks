```python
#coding:cp932
# 2021.03.31 14.47 Created

import numpy as np
import pandas as pd
import copy
from scipy import stats as st
import matplotlib.pyplot as pyplot

def main():
    #print("Enter man()")
    #data = pd.read_csv("./archive/StudentsPerformance.csv")
    data = pd.read_csv("/kaggle/input/students-performance-in-exams/StudentsPerformance.csv")
    columns = data.columns
    #print(columns)
    #print(data[columns[5]])

   # display scatter chart
    pyplot.figure()
    
    for i in range(0,3):
        for j in range(0,3):
            #print("row=",columns[5+i], "column=", columns[5+j])
            pyplot.subplot(3,3,i*3+j+1)
            pyplot.scatter(data[columns[5+i]],data[columns[5+j]])
            pyplot.xlabel(data.columns.values[5+i])
            pyplot.ylabel(data.columns.values[5+j])
    
    pyplot.show()
    
    # coefficient between parameters
    print("Coefficient between parameters")
    s = [data[columns[5]],data[columns[6]],data[columns[7]]]
    print(np.corrcoef(s))

    pivot = data.pivot_table(index='gender', values=['math score', 'reading score', 'writing score'])
    #print(pivot)
    male  = data[data['gender'] == 'male']
    female = data[data['gender'] == 'female']
    
    # stacked hist chart with gender 
    pyplot.figure()
    
    for i in range(0,3):
        pyplot.subplot(1,3,i+1)
        pyplot.hist([male[male.columns[5+i]], female[female.columns[5+i]]], stacked=True, label=['male','female'] )
        pyplot.legend()
        pyplot.xlabel(data.columns[5+i])
        pyplot.ylabel("count")
        
    pyplot.show()

    # any diff. with gendaer?
    # math
    print("Gender difference")
    #print(data[data['gender']=='female'][columns[0]], ", " ,data[data['gender']=='female'][columns[5]])
    f, p = st.f_oneway(data[data['gender']=='female'][columns[5]], data[data['gender']=='male'][columns[5]])
    print("math score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("math score (gender diff.)")
    ax1.boxplot([data[data['gender']=='female'][columns[5]], data[data['gender']=='male'][columns[5]]], labels=['female', 'male'])
    pyplot.show()
    # reading
    #print(data[data['gender']=='female'][columns[0]], ", " ,data[data['gender']=='female'][columns[5]])
    f, p = st.f_oneway(data[data['gender']=='female'][columns[6]], data[data['gender']=='male'][columns[6]])
    print("reading score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("reading score (gender diff.)")
    ax1.boxplot([data[data['gender']=='female'][columns[6]], data[data['gender']=='male'][columns[6]]], labels=['female', 'male'])
    pyplot.show()
    # writing
    #print(data[data['gender']=='female'][columns[0]], ", " ,data[data['gender']=='female'][columns[5]])
    f, p = st.f_oneway(data[data['gender']=='female'][columns[7]], data[data['gender']=='male'][columns[7]])
    print("writing score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("writing score (gender diff.)")
    ax1.boxplot([data[data['gender']=='female'][columns[7]], data[data['gender']=='male'][columns[7]]], labels=['female', 'male'])
    pyplot.show()

    # race diff.
    # math
    print("Race difference")
    f, p = st.f_oneway(data[data['race/ethnicity']=='group A'][columns[5]], 
                       data[data['race/ethnicity']=='group B'][columns[5]],
                       data[data['race/ethnicity']=='group C'][columns[5]],
                       data[data['race/ethnicity']=='group D'][columns[5]],
                       data[data['race/ethnicity']=='group E'][columns[5]])
    print("math score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("math score (race diff.)")
    ax1.boxplot([data[data['race/ethnicity']=='group A'][columns[5]], 
                 data[data['race/ethnicity']=='group B'][columns[5]],
                 data[data['race/ethnicity']=='group C'][columns[5]],
                 data[data['race/ethnicity']=='group D'][columns[5]],
                 data[data['race/ethnicity']=='group E'][columns[5]]
                 ], labels=['group A', 'group B', 'group C', 'group D', 'group E'])
    pyplot.show()

    # reading
    f, p = st.f_oneway(data[data['race/ethnicity']=='group A'][columns[6]], 
                       data[data['race/ethnicity']=='group B'][columns[6]],
                       data[data['race/ethnicity']=='group C'][columns[6]],
                       data[data['race/ethnicity']=='group D'][columns[6]],
                       data[data['race/ethnicity']=='group E'][columns[6]])
    print("reading score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("reading score (race diff.)")
    ax1.boxplot([data[data['race/ethnicity']=='group A'][columns[6]], 
                 data[data['race/ethnicity']=='group B'][columns[6]],
                 data[data['race/ethnicity']=='group C'][columns[6]],
                 data[data['race/ethnicity']=='group D'][columns[6]],
                 data[data['race/ethnicity']=='group E'][columns[6]]
                 ], labels=['group A', 'group B', 'group C', 'group D', 'group E'])
    pyplot.show()

    # writing
    f, p = st.f_oneway(data[data['race/ethnicity']=='group A'][columns[7]], 
                       data[data['race/ethnicity']=='group B'][columns[7]],
                       data[data['race/ethnicity']=='group C'][columns[7]],
                       data[data['race/ethnicity']=='group D'][columns[7]],
                       data[data['race/ethnicity']=='group E'][columns[7]])
    print("writing score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("writing score  (race diff.)")
    ax1.boxplot([data[data['race/ethnicity']=='group A'][columns[7]], 
                 data[data['race/ethnicity']=='group B'][columns[7]],
                 data[data['race/ethnicity']=='group C'][columns[7]],
                 data[data['race/ethnicity']=='group D'][columns[7]],
                 data[data['race/ethnicity']=='group E'][columns[7]]
                 ], labels=['group A', 'group B', 'group C', 'group D', 'group E'])
    pyplot.show()

    # education diff.
    # math
    print("parental level of education difference")
    f, p = st.f_oneway(data[data['parental level of education']=="associate's degree"][columns[5]], 
                       data[data['parental level of education']=="bachelor's degree"][columns[5]],
                       data[data['parental level of education']=="high school"][columns[5]],
                       data[data['parental level of education']=="master's degree"][columns[5]],
                       data[data['parental level of education']=="some college"][columns[5]],
                       data[data['parental level of education']=="some high school"][columns[5]])
    print("math score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("math score (education diff.)")
    ax1.boxplot([data[data['parental level of education']=="associate's degree"][columns[5]], 
                 data[data['parental level of education']=="bachelor's degree"][columns[5]],
                 data[data['parental level of education']=="high school"][columns[5]],
                 data[data['parental level of education']=="master's degree"][columns[5]],
                 data[data['parental level of education']=="some college"][columns[5]],
                 data[data['parental level of education']=="some high school"][columns[5]]
                 ], labels=["associate's degree", "bachelor's degree", "high school", "master's degree", "some college", "ome high school"])
    pyplot.show()
    
    # reading
    f, p = st.f_oneway(data[data['parental level of education']=="associate's degree"][columns[6]], 
                       data[data['parental level of education']=="bachelor's degree"][columns[6]],
                       data[data['parental level of education']=="high school"][columns[6]],
                       data[data['parental level of education']=="master's degree"][columns[6]],
                       data[data['parental level of education']=="some college"][columns[6]],
                       data[data['parental level of education']=="some high school"][columns[6]])
    print("reading score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("reading score  (education diff.)")
    ax1.boxplot([data[data['parental level of education']=="associate's degree"][columns[6]], 
                 data[data['parental level of education']=="bachelor's degree"][columns[6]],
                 data[data['parental level of education']=="high school"][columns[6]],
                 data[data['parental level of education']=="master's degree"][columns[6]],
                 data[data['parental level of education']=="some college"][columns[6]],
                 data[data['parental level of education']=="some high school"][columns[6]]
                 ], labels=["associate's degree", "bachelor's degree", "high school", "master's degree", "some college", "ome high school"])
    pyplot.show()
    
    # writing
    f, p = st.f_oneway(data[data['parental level of education']=="associate's degree"][columns[7]], 
                       data[data['parental level of education']=="bachelor's degree"][columns[7]],
                       data[data['parental level of education']=="high school"][columns[7]],
                       data[data['parental level of education']=="master's degree"][columns[7]],
                       data[data['parental level of education']=="some college"][columns[7]],
                       data[data['parental level of education']=="some high school"][columns[7]])
    print("writing score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("writing score  (education diff.)")
    ax1.boxplot([data[data['parental level of education']=="associate's degree"][columns[7]], 
                 data[data['parental level of education']=="bachelor's degree"][columns[7]],
                 data[data['parental level of education']=="high school"][columns[7]],
                 data[data['parental level of education']=="master's degree"][columns[7]],
                 data[data['parental level of education']=="some college"][columns[7]],
                 data[data['parental level of education']=="some high school"][columns[7]]
                 ], labels=["associate's degree", "bachelor's degree", "high school", "master's degree", "some college", "ome high school"])
    pyplot.show()

    #lunch diff.
    # math
    print("parental level of lunch difference")
    f, p = st.f_oneway(data[data['lunch']=="standard"][columns[5]], 
                       data[data['lunch']=="free/reduced"][columns[5]])
    print("math score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("math score (lunch diff.)")
    ax1.boxplot([data[data['lunch']=="standard"][columns[5]], 
                 data[data['lunch']=="free/reduced"][columns[5]]
                 ], labels=["standard","free/reduced"])
    pyplot.show()
    
    # reading
    print("parental level of lunch difference")
    f, p = st.f_oneway(data[data['lunch']=="standard"][columns[6]], 
                       data[data['lunch']=="free/reduced"][columns[6]])
    print("reading score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("reading score (lunch diff.)")
    ax1.boxplot([data[data['lunch']=="standard"][columns[6]], 
                 data[data['lunch']=="free/reduced"][columns[6]]
                 ], labels=["standard","free/reduced"])
    pyplot.show()
    
    # writing
    print("parental level of lunch difference")
    f, p = st.f_oneway(data[data['lunch']=="standard"][columns[7]], 
                       data[data['lunch']=="free/reduced"][columns[7]])
    print("writing score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("writing score (lunch diff.)")
    ax1.boxplot([data[data['lunch']=="standard"][columns[7]], 
                 data[data['lunch']=="free/reduced"][columns[7]]
                 ], labels=["standard","free/reduced"])
    pyplot.show()


    #test preparation
    # math
    print("parental level of test preparation course difference")
    f, p = st.f_oneway(data[data['test preparation course']=="none"][columns[5]], 
                       data[data['test preparation course']=="completed"][columns[5]])
    print("math score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("math score (test preparation diff.)")
    ax1.boxplot([data[data['test preparation course']=="none"][columns[5]], 
                 data[data['test preparation course']=="completed"][columns[5]]
                 ], labels=["none","completed"])
    pyplot.show()
    
    # reading
    print("parental level of test preparation course difference")
    f, p = st.f_oneway(data[data['test preparation course']=="none"][columns[6]], 
                       data[data['test preparation course']=="completed"][columns[6]])
    print("reading score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("reading score (test preparation diff.)")
    ax1.boxplot([data[data['test preparation course']=="none"][columns[6]], 
                 data[data['test preparation course']=="completed"][columns[6]]
                 ], labels=["none","completed"])
    pyplot.show()
    
    # writing
    print("parental level of test preparation course difference")
    f, p = st.f_oneway(data[data['test preparation course']=="none"][columns[7]], 
                       data[data['test preparation course']=="completed"][columns[7]])
    print("test preparation score F=%f, p-val=%f", (f, p))
    fig, ax1 = pyplot.subplots()
    ax1.set_title("writing score (test preparation diff.)")
    ax1.boxplot([data[data['test preparation course']=="none"][columns[7]], 
                 data[data['test preparation course']=="completed"][columns[7]]
                 ], labels=["none","completed"])
    pyplot.show()

if __name__ == "__main__":
    main()
```
