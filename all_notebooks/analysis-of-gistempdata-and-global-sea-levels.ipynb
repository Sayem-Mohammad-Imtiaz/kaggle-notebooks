```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 1.5, 1.5])
title = ax1.set_title("Global temperature from 1880 to 2000", fontsize='large')
ax1.plot(data['Year'], data['SHem'], 'r', label='Southern Hemisphere')
ax1.plot(data['Year'], data['NHem'], 'b', label='Northern Hemisphere')
ax1.plot(data['Year'], data['Glob'], 'y--', label='Global')
ax1.set_xticks(np.arange(1880, 2020, step=10))
ax1.legend()

plt.xlabel('Year', fontsize='large')
plt.ylabel('Deviation', fontsize='large')
```

```python
print(data['Glob'].min())
print(data['Glob'].max())
```

```python
fig = plt.figure(figsize=(20,20))
ax1 = plt.subplot(111, projection='polar')

theta = np.linspace(0, 2*np.pi, 10)

ax1.axes.get_yaxis().set_ticklabels([])
ax1.axes.get_xaxis().set_ticklabels([])

fig.set_facecolor("#000100")

full_circle_thetas = np.linspace(0, 2*np.pi, 1000)

for i in range(3,13):
    r = data['Glob'][(10*i):((10*i)+10)]
    line_radi = [r.mean()]*1000
    ax1.plot(theta, r, marker = 'o')
    ax1.plot(full_circle_thetas, line_radi, c='hotpink')
    ax1.text(np.pi/2, r.mean() +.5, "Average difference: " + str(data['Year'].iloc[10*i]), color="hotpink", ha='center', fontdict={'fontsize': 10})
    
plt.show()
```

Since this is jsut a bit congested lets try a different approach. Lets look specifically at the north and south poles.

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 1.5, 1.5])
title = ax1.set_title("Global temperature from 1880 to 2000, South Pole", fontsize='large')
ax1.plot(data['Year'], data['90S-64S'], 'r', label='South Pole')
ax1.set_xticks(np.arange(1880, 2020, step=10))
ax1.legend()

x_new = np.linspace(1880, 2010, 15)
a_BSpline = make_interp_spline(data['Year'], data['90S-64S'])
y_new = a_BSpline(x_new)
plt.plot(x_new, y_new)

m, b = np.polyfit(data['Year'], data['90S-64S'], 1)
plt.plot(data['Year'], m*data['Year'] + b)

plt.xlabel('Year', fontsize='large')
plt.ylabel('Deviation', fontsize='large')
```

```python
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 1.5, 1.5])
title = ax1.set_title("Global temperature from 1880 to 2000, NorthPole", fontsize='large')
ax1.plot(data['Year'], data['64N-90N'], 'r', label='North Pole')
ax1.set_xticks(np.arange(1880, 2020, step=10))
ax1.legend()

x_new = np.linspace(1880, 2010, 15)
a_BSpline = make_interp_spline(data['Year'], data['64N-90N'])
y_new = a_BSpline(x_new)
plt.plot(x_new, y_new,label='20 year intervals')

m, b = np.polyfit(data['Year'], data['64N-90N'], 1)
plt.plot(data['Year'], m*data['Year'] + b)

plt.xlabel('Year', fontsize='large')
plt.ylabel('Deviation', fontsize='large')
```

## Solar Irradiation

![](https://climate.nasa.gov/system/content_pages/main_images/1_Greenhouse_effect_revised_English.gif)

```python
solar_data = pd.read_csv('/kaggle/input/solar-irradiation/solar_irra.csv')
solar_data.head()
```

```python
solar_data['Total Solar Irradiance (TSI) at 1-AU, W/m^2'].count()
```

```python
solar_non_zero = solar_data[solar_data['Total Solar Irradiance (TSI) at 1-AU, W/m^2'] != 0]
for row in solar_non_zero:
    solar_non_zero['Year'] = solar_non_zero['Nominal Data Time, YYYYMMDD'].astype(str).str[:4]
solar_non_zero.head()
```

```python
plt.plot(range(2003,2021), solar_non_zero.groupby(['Year']).mean(['Total Solar Irradiance (TSI) at 1-AU, W/m^2']))
plt.xlabel('Nominal Data Time')
plt.ylabel('Total Solar Irradiance (TSI)')
```
