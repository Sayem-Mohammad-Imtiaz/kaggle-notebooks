```python
import pyspark
from pyspark.sql import SparkSession
from pyspark.sql import functions as f

spark = SparkSession.builder.appName("PySparkKaggle").getOrCreate()
```

```python
moviesDF = spark.read.option("header", True).csv("../input/msis405l21cttt-spring-2021-assignment-7/movies_small.csv")
```

# **Câu 1**

```python
moviesDF.filter(f.array_contains(f.split('genres','\|'),'Action')).show(5, truncate=False)
```

# **Câu 2**

```python
moviesDF2 = moviesDF.withColumn('genre_array', f.split('genres', '\|')).withColumn('count', f.size('genre_array'))
moviesDF2.select('movieId', 'title', 'count').show(5, truncate=False)
```

# **Câu 3**

```python
moviesDF.withColumn('arr', f.split('genres', '\|')).withColumn('genres', f.explode('arr')).groupBy('genres').count().sort('count').show(5)
```

# **Câu 4**

```python
moviesDF4 = moviesDF.withColumn('genre_array', f.split('genres', '\|')).withColumn('genre', f.explode('genre_array'))
moviesDF4.groupBy('genre').agg(f.collect_list('title').alias('movie_list')).show(truncate=100)
```

# **Câu 5**

```python
moviesDF.withColumn('arr', f.split('title', ' ')).withColumn('last_element', f.element_at('arr', -1)).withColumn('year', f.substring('last_element', 2,4).cast('int')).withColumn('arr_genres', f.split('genres', '\|')).withColumn('genres', f.explode('arr_genres')).groupBy('genres').agg(f.min('year').alias('first_appearance')).filter(f.col('genres').isin(['Animation','Sci-Fi'])).show(10, truncate=False)
```
