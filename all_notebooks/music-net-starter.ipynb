```python
# Let's try to open one of the file (Beethoven ??)
mid=MidiFile('../input/musicnet-dataset/musicnet_midis/musicnet_midis/Beethoven/2313_qt15_1.mid',clip=True)
mid
```

## This means this is a midi file from the above location and type 1 which means all the tracks starts at the same time and 6 tracks here means 6 synchronous tracks while showing 11947 messages in total
## There are 3 types in total :*
## 1) 0 Type (Single Track) : all messages are saved in one track
## 2) 1 Type (Synchronous Track) : all tracks start at the same time
## 3) 2 Type (Asynchronous Track) :  each track is independent of the others

```python
for track in mid.tracks:
    print(track)
```

## From this we can see the composition of messages shown by the different tracks which are synchronously working with each other for e.g. :
## Violin1 has 3277 messages in total

```python
for msg in mid.tracks[1]:
    print(msg)
```

## These are the messages of Violin1 these all will account for total number of 3277

## From the above example we can see that there are some repetitions of the messages we can clean that up using a code

```python
message_numbers = []
duplicates = []

for track in mid.tracks:
    if len(track) in message_numbers:
        duplicates.append(track)
    else:
        message_numbers.append(len(track))

for track in duplicates:
    mid.tracks.remove(track)

mid.save('new_song.mid')
```

```python
# Using Librosa to have a look at the data
import librosa
x, sr = librosa.load('../input/musicnet-dataset/musicnet/musicnet/train_data/1727.wav')
```

```python
print(x.shape)
print(sr)
```

## This is a huuuge file as we can see from the shape of the array

## Visualizing the data/Audio File.

```python
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x, sr=sr)
```

# Displaying a Spectrogram

```python
X = librosa.stft(x)
Xdb = librosa.amplitude_to_db(abs(X))
plt.figure(figsize=(14, 5))
librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')
```
