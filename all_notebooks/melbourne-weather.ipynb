```python
# Import Libs
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# Load Data File
temperature_df = pd.read_csv('../input/melbourneweather/MelbourneWeather.csv')
temperature_df.reset_index(inplace=True)
```

```python
# Create a Combo SNS Chart
fig, ax1 = plt.subplots(figsize=(26,14))
# Set Title for Chart
ax1.set_title('Yearly Melbourne Weather - Temperatures Over Time', fontsize=16)
# Title for X-Axis
ax1.set_xlabel('Day of Year', fontsize=16)
# Title for Left Side Y-Axis
ax1.set_ylabel('Maximum Temperature', fontsize=16)

# Plot Chart for Maximum Temperature
ax1 = sns.lineplot(
    x="Day",
    y="Maximum",
    hue='Year',
    palette="Reds",
    data=temperature_df)

# Define Range for Y-Axis
ax1.set(ylim=(-2, 50))

# Enable Dual Axis Mode
ax2 = ax1.twinx()

# Title for Left Side Y-Axis
ax2.set_ylabel('Minimum Temperature', fontsize=16)

# Plot Chart for Minimum Temperature
ax2 = sns.lineplot(
    x="Day",
    y="Minimum",
    hue='Year',
    palette="Blues",
    data=temperature_df)


ax2.set_ylim(ax1.get_ylim())
# Add Lines for Months
xposition = [0,30,60,90,120,150,180,211,241,270,300,333,365]
for xc in xposition:
    plt.axvline(x=xc, color='grey', linestyle='--')
plt.show()

# Display Combo Chart
plt.show()
```
