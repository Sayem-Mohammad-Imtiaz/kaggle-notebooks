```python
# Assigning posts table to dataframes
posts_df = pd.read_csv('/kaggle/input/relational-data-engineering/posts_table.csv')
posts_df.head(5)
```

```python
# Assigning reactions table to dataframes
reactions_df = pd.read_csv('/kaggle/input/relational-data-engineering/reactions_table.csv')
reactions_df.head(5)
```

```python
# Adding user ids to the user table:
user_ids = list(range(1, 1001))
user_df['User'] = user_ids  # column User is added
user_df.head()
```

```python
# merging tables of posts and reactions to get total post and reactions by each user
post_reaction_df = posts_df.merge(reactions_df[['User', 'Reaction Type']], left_on='User', right_on='User', how='outer')[['User', 'Post Type', 'Reaction Type']]
```

```python
# Top five Users with most posts and reactions combined
highest_posts_reactions = post_reaction_df.groupby(['User']).size().to_frame('Posts and Reactions').reset_index().sort_values(['Posts and Reactions'], ascending=[False])
top_five = highest_posts_reactions.head()
top_five
```

```python
# Adding Name and Surname to the result table from last cell
top_five_names = top_five.merge(user_df[['Full_Name', 'User']], left_on='User', right_on='User', how='inner')
top_five_names
```

```python
# Ans. of Question 02: Top five people with most post & reaction combined are Ali Mueller, Zoe Picard, Agaba Gwahsi, Robert Di Lillo, Hans Bond.
```

```python
from matplotlib import pyplot as plt
import networkx as nx
```

```python
# Assigning friends tables to dataframes
friends_df = pd.read_csv('/kaggle/input/relational-data-engineering/friends_table.csv')
friends_df.head(5)
```

```python
#Filter out all "Jean-Luc Mueller" from user data.

filter = (user_df["Full_Name"] == 'Jean-Luc Mueller')
user_JeanLuc_Mueller = user_df[filter]
user_JeanLuc_Mueller
```

```python
# Merging user_JeanLuc_Mueller with friends dataframe to find Jean-Luc Mueller friends ID

Friendship_df= pd.merge(user_JeanLuc_Mueller,friends_df, left_on='User', right_on="Friend 1",how="left")
Friendship_df
```

```python
Friends_ID_of_JeanLuc_Mueller = Friendship_df['Friend 2']
Friends_ID_of_JeanLuc_Mueller
```

```python
# Finding the friends name from user table and adding user with that table

friends_of_JeanLuc_Mueller = user_df.loc[Friends_ID_of_JeanLuc_Mueller, :]
JeanLuc_Mueller_and_his_first_degree_friend = user_JeanLuc_Mueller.append(friends_of_JeanLuc_Mueller)
JeanLuc_Mueller_and_his_first_degree_friend
```

```python
# Finding the second degree friends of Jean-Luc Mueller

JeanLuc_Mueller_and_his_second_degree_friend = JeanLuc_Mueller_and_his_first_degree_friend.merge(friends_df, left_on='User', right_on="Friend 2", how="left")
JeanLuc_Mueller_and_his_second_degree_friend_df = JeanLuc_Mueller_and_his_second_degree_friend.merge(user_df, left_on='User', right_on="User", how="left")
JeanLuc_Mueller_and_his_second_degree_friend_df   # This is the dataframe with second degree friend of Jean-Luc Mueller
```

```python
# ------- 2nd degree DIRECTED friendship Graph for all users with name "Jean-Luc Mueller"
 

# Building my graph. Note that I use the DiGraph function to create the Directed graph!
G=nx.from_pandas_edgelist(JeanLuc_Mueller_and_his_second_degree_friend_df, 'Full_Name_x', 'Full_Name_y', create_using=nx.DiGraph() )

# Make the graph
nx.draw(G, with_labels=True, node_size=200, alpha=0.5, arrows = True, font_weigth='bold')
```
