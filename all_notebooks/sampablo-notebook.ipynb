# Shape Method
To know the number of columns and rows we use **shape** method.

The left value is for rows, and the right value is for columns.

```python
print("Data train: ",data_train.shape)
print("Data gender: ",data_gender.shape)
print("Data test: ",data_test.shape)
print("Data gender json: ",gender_json.shape)
```

We have a option if we dont want to count duplicate cells, using **drop_duplicates()** method.

In this example we are going to duplicate temp_gender data to deleted after we use shape method.

```python
temporal_gender = data_gender.append(temp_gender)
#original shape
print("Original shape: ",data_gender.shape)

#with duplicated columns
print("Duplicated shape: ", temporal_gender.shape)

#removing duplicated columns
temp = temporal_gender.drop_duplicates()
print("New shape after delete duplicates: ", temp.shape)
```
