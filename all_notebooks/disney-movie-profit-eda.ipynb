```python
df = pd.read_csv('/kaggle/input/disney-movies-19372016-total-gross/disney_movies_total_gross.csv', parse_dates=['release_date'])
df.head()
```

```python
gross_adj = df.sort_values(by = 'inflation_adjusted_gross', ascending = False)
gross_adj.head()
```

```python
#plot which movies net highest profit
gross_adj10 = gross_adj.head(10)
plt.figure(figsize=(10,10))
sns.barplot(x=gross_adj10.movie_title,
            y=gross_adj10.inflation_adjusted_gross, 
           palette="Paired")
plt.xlabel("Movie", size = 15)
plt.ylabel("Gross Income (billions)", size = 15)
plt.title("Disney Movie Incomes Adjusted for Inflation", size = 18)
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

```python
genre = df.groupby(['genre']).mean()
genre1 = genre.reset_index()
genre1['inflation_adjusted_gross'] = genre1['inflation_adjusted_gross'].astype(int)
genre1 = genre1.sort_values(by = 'inflation_adjusted_gross', ascending = False)
genre1.head(10)
```

```python
#plot which genres net highest profit
gross_adj10 = gross_adj.head(10)
plt.figure(figsize=(10,10))
sns.barplot(x=genre1.genre,
            y=genre1.inflation_adjusted_gross, 
            palette="Paired")
plt.xlabel("Genre", size = 15)
plt.ylabel("Gross Income (billions)", size = 15)
plt.title("Disney Genres Adjusted for Inflation", size = 18)
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

```python
df['year'] = pd.DatetimeIndex(df['release_date']).year
genrey = df.groupby(['genre', 'year']).mean()
genre2 = genrey.reset_index()
genre2['inflation_adjusted_gross'] = genre2['inflation_adjusted_gross'].astype(int)
genre2 = genre2.sort_values(by = 'inflation_adjusted_gross', ascending = False)
genre2.head(10)
```

```python
#Plot Time Series for gross income over time
plt.figure(figsize=(10,10))
sns.lineplot(x=genre2.year,
            y=genre2.inflation_adjusted_gross)
plt.xlabel("Year", size = 15)
plt.ylabel("Gross Income (billions)", size = 15)
plt.title("Disney Movie Profits Over Time Adjusted for Inflation", size = 18)
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```

```python
#plot how genre profit changes over time
gross_adj10 = gross_adj.head(10)
plt.figure(figsize=(10,10))
sns.lineplot(x=genre2.year,
            y=genre2.inflation_adjusted_gross, 
             hue=genre2.genre)
plt.xlabel("Year", size = 15)
plt.ylabel("Gross Income (billions)", size = 15)
plt.title("Disney Genre Profits Over Time Adjusted for Inflation", size = 18)
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
```
