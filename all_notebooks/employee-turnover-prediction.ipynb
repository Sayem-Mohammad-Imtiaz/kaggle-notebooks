```python
from pycaret.classification import *
```

```python
exp_clf102 = setup(data = data, target = 'Attrition', session_id=123,
                  normalize = True, 
                  transformation = True, 
                  ignore_low_variance = True,
                  remove_multicollinearity = True, multicollinearity_threshold = 0.95,
                  log_experiment = True, experiment_name = 'turnover')
```

```python
top3 = compare_models(n_select = 3)
```

```python
lr=create_model('lr')
```

```python
print(lr)
```

```python

tuned_lr = tune_model(lr)
```

```python
print(tuned_lr)
```

```python
plot_model(tuned_lr, plot = 'auc')
```

```python
plot_model(tuned_lr, plot = 'pr')
```

```python
plot_model(tuned_lr, plot='feature')
```

```python
plot_model(tuned_lr, plot = 'confusion_matrix')
```

```python
evaluate_model(tuned_lr)
```

```python
final_lr = finalize_model(tuned_lr)
```

```python
print(final_lr)
```

```python
predictions = predict_model(final_lr, data=data)
predictions.head(20)
```

```python
from pycaret.utils import check_metric
check_metric(predictions['Attrition'], predictions['Label'], metric = 'Accuracy')
```
