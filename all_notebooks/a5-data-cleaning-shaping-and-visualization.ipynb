The dataset I'm interested in is the leading causes of deaths in 2019 provided by the CDC offcial website. This is interesting to me because I want to know what are the top causes of death. I want to explore the differences of death causes among age groups and understand what is the most common cause of death across all age groups.

**Step 1**: Load data using panda and view basic data structure

```python
data = pd.read_csv('../input/cdccauses/lcd.csv')
data.head()
```

**Step 2:** Use a histogram to show the sum of deaths for each cause of deaths for all age groups.

```python
# import plotly to show hover over effect
# Show x axis as the cause of death and y axis the sum of deaths
import plotly.express as px
hist = px.histogram(data, x="Cause of Death", y="Deaths")
hist.show()
```

Step 3: Use a scatter plot graph to view the number of deaths per each cause of death and how they distributed across each age group.

```python
#start a new figure and make the x axis the age group and y axis the number of deaths. 
#To hoverover the graph, can view the number of deaths and the ranking of the specific cause of death for a age group
fig = px.scatter(data, x=data['Age Group'], y=data['Deaths'] , hover_name="Cause of Death", hover_data=["Cause of Death", "Rank"])
fig.show()
```
