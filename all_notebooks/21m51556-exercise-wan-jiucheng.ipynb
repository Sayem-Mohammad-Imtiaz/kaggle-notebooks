# Country selected
## The country I selected here is Chile

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

selected_country_1='Chile'
selected_country_2='Chile'

df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1 = df1[df1['country']==selected_country_1]
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2 = df2[df2['country']==selected_country_2]
df1.index = pd.to_datetime(df1['date'])
df2.index = pd.to_datetime(df2['date'])
```

# Figure
## Here is the figure of the daily new cases and daily vaccinations in Chile.

```python
fig,ax = plt.subplots(figsize=(10,5))
ax.plot(df1.index,df1['daily_new_cases'],label='cases')
ax1 = ax.twinx()
ax1.plot(df2.index,df2['daily_vaccinations'],'r',label='vaccinations')
ax1.legend()
ax.legend()
plt.show()
```

# Analysis
## After improve visualization, we can see in February of 2021, the cases is stopped with the rising of the vaccinations, but in April, The number of cases rises with the number of the vaccinations.
## By looking up the information online, I got that the country's rapid vaccination campaign gave the people a false sense of security and contributed to a sharp rise in new infections and deaths.
## The government eased restrictions “excessively” after imposing very strict quarantines.
## Here is the news and reports
## https://www.aa.com.tr/en/americas/why-chile-breaks-vaccination-and-infections-records/2202023

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_new_cases = go.Scatter(x=df1.index,y=df1['daily_new_cases'],yaxis='y1',name='Daily confirmed cases')
daily_vaccinations = go.Scatter(x=df2.index,y=df2['daily_vaccinations'],yaxis='y2',name='Daily vaccinations')

layout_obj = go.Layout(title='COVID vs. Vaccines',xaxis=dict(title='Date'),yaxis=dict(title='Daily cases'),yaxis2=dict(title='Vaccinations',side='right',overlaying='y'))
fig = go.Figure(data=[daily_new_cases,daily_vaccinations],layout=layout_obj)
iplot(fig)
```

# Ranking 1
## Here is the global ranking of that country in terms of the total number of cases

```python
df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
df1 = df1.groupby('country').sum()['daily_new_cases'].sort_values(ascending=False)
print(df1.to_string())
```

# Ranking 2
## Here is the global ranking of that country in terms of the total vaccinations per number of people

```python
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df2 = df2.groupby('country').max()['people_fully_vaccinated_per_hundred'].sort_values(ascending=False)
print(df2.to_string())
```
