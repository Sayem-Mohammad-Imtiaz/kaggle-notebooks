## Predicting Liver Disease in Patients  

Given *medical data collected from patients northeast of Andhra Pradesh*, let's try to predict whether a given patient has **liver disease** or not.

We will use a TensorFlow ANN to make our predictions.

# Getting Started

```python
import numpy as np
import pandas as pd
import plotly.express as px

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

import tensorflow as tf
```

```python
data = pd.read_csv('../input/indian-liver-patient-records/indian_liver_patient.csv')
```

```python
data
```

```python
data.info()
```

# Preprocessing

## Missing Values

```python
data.isna().sum()
```

```python
data['Albumin_and_Globulin_Ratio'] = data['Albumin_and_Globulin_Ratio'].fillna(data['Albumin_and_Globulin_Ratio'].mean())
```

## Encoding

```python
def binary_encode(df, column, positive_value):
    df = df.copy()
    df[column] = df[column].apply(lambda x: 1 if x == positive_value else 0)
    return df
```

```python
data = binary_encode(data, 'Gender', 'Male')
```

Let's change the labels to 0, 1 instead of 1, 2

```python
data = binary_encode(data, 'Dataset', 1)
```

## Splitting and Scaling

```python
data
```

```python
y = data['Dataset']
X = data.drop('Dataset', axis=1)
```

```python
scaler = StandardScaler()

X = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8)
```

# Training

```python
X.shape
```

```python
y.sum() / len(y)
```

```python
inputs = tf.keras.Input(shape=(10,))
x = tf.keras.layers.Dense(64, activation='relu')(inputs)
x = tf.keras.layers.Dense(64, activation='relu')(x)
outputs = tf.keras.layers.Dense(1, activation='sigmoid')(x)

model = tf.keras.Model(inputs=inputs, outputs=outputs)


model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=[
        'accuracy',
        tf.keras.metrics.AUC(name='auc')
    ]
)


batch_size = 64
epochs = 31

history = model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    batch_size=batch_size,
    epochs=epochs,
    verbose=0
)
```

# Results

```python
fig = px.line(
    history.history,
    y=['loss', 'val_loss'],
    labels={'index': "Epoch", 'value': "Loss"},
    title="Training and Validation Loss"
)

fig.show()
```

```python
np.argmin(history.history['val_loss']) + 1
```

```python
model.evaluate(X_test, y_test)
```
