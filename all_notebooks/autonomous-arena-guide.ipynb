# Introduction

The dataset was collected on a manually-built indoor car track, using a Raspberry Pi controlled vehicle. The images were obtained from a camera mounted in front of the car that captured the frames at 8fps. A self-built android app was used to control the car through a Bluetooth module. There were 4 major signals to maneuver the car viz. front (F), left (L), right (R), and stop (S).

# Read data

````python
# read csv file
trackData = pd.read_csv("/kaggle/input/indoor-car-track/autonomous_arena.csv")
trackData.head()
````

````python
# split images and labels
label = trackData["448"]
image = trackData.drop("448", axis=1)
````

# Stats

````python
print("Total samples: {}".format(image.shape[0]))
print("Image size: 14x32")
print("Label distribution:")
print(label.value_counts())
````

# Visualization

````python
plt.figure(figsize=(20,5))
plot = 1
for i in range(2):
    for j in range(5):
        idx = random.randint(1, image.shape[0])
        img = image.iloc[idx].values.reshape((14,32))
        
        plt.subplot(2, 5, plot)
        plt.imshow(img, cmap="gray")
        plt.title("Label: {}".format(label.iloc[idx]))
        
        plot += 1
plt.show()
````
