```python
df['Serving_Size_gm']=df['Serving Size'].apply(lambda x: x.split(' (')[-1])
df['Serving_Size_gm']=df['Serving_Size_gm'].apply(lambda x: x.split(' ')[0])
df['Serving_Size_gm']
```

```python
df.head()
```

```python
df.drop('Serving Size',1, inplace=True)
```

```python
df['Serving_Size_gm']=df['Serving_Size_gm'].astype('float64')
```

```python
num_cols=[x for x in df.columns if df[x].dtypes!='O']
print(num_cols)
```

```python
count=pd.DataFrame(df['Category'].value_counts())
count=count.rename_axis(['category']).reset_index()
count['count']=count['Category']
count.drop('Category',1, inplace=True)
count
px.bar(count,x='category', y='count')
```
