```python
X = bs [["Name", "Author", "Reviews", "Price", "Year", "Genre"]]
X
```

```python
y = bs[["User Rating"]]
y
```

```python
X_cat = X[["Genre"]]
X_num = X[["Reviews", "Price", "Year"]]
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
X_cat_OH =  encoder.fit_transform(X_cat)
X_cat_OH
```

```python
X_cat1 = pd.DataFrame(X_cat_OH.toarray())
X_cat1
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X1=X[["Name"]].Name.tolist()
X1 = vectorizer.fit_transform(X1)
X1=pd.DataFrame(X1.toarray())
X1.head()
```

```python
bs = [X1, X_cat1, X_num]
```

```python
bs1 = pd.concat(bs, axis=1).reset_index(drop = True)
bs1.head
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(bs1, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model1=GradientBoostingRegressor(n_estimators=1000,max_features="log2",max_depth=20,learning_rate=0.09,
                               random_state=0).fit(X_train,np.asarray(y_train).ravel())
model1
```

```python
from sklearn.metrics import r2_score
train_pred=model1.predict(X_train)
test_pred=model1.predict(X_test)
print("R-squared on train data: ",r2_score(y_train,train_pred))
print("R-squared on test data: ",r2_score(y_test,test_pred))
```

```python
from sklearn.metrics import mean_squared_error as mse
train_mse=mse(y_train,train_pred)
test_mse=mse(y_test,test_pred)
print("Root mean squared error on train data: ",np.sqrt(train_mse))
print("Root mean squared error on test data: ",np.sqrt(test_mse))
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(X_train, np.asarray(y_train).ravel())
y_pred = linreg.predict(X_test)
```

```python
train_pred1=linreg.predict(X_train)
print("R-squared on train data: ",r2_score(y_train,train_pred1))
print("R-squared on test data: ",r2_score(y_test,y_pred))
```

```python
train_mse=mse(y_train,train_pred1)
test_mse=mse(y_test,y_pred)
print("Root mean squared error on train data: ",np.sqrt(train_mse))
print("Root mean squared error on test data: ",np.sqrt(test_mse))
```
