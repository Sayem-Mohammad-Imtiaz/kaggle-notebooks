```python
import matplotlib.pyplot as plt

Company = ['TOP Media', 'BG', 'MBK', 'Choeun', 'JYP']
Members = [4, 14, 9, 8, 4]

plt.plot(Company, Members, color='red', marker='o')
plt.title('Company Vs Members', fontsize=14)
plt.xlabel('Company', fontsize=14)
plt.ylabel('Members', fontsize=14)
plt.grid(True)
plt.show()
```

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np

fig = plt.figure()
ax = fig.add_axes([0,0,3.5,1])

belajar = pd.read_csv('../input/kpopdb/kpop_idols_boy_groups.csv')
belajar.head()

x_label = belajar['Company']
plt.bar(x=np.arange(len(x_label)),height=belajar['Members'])
plt.xticks(np.arange(len(x_label)),belajar['Company'], rotation=90)
plt.xlabel('Nama Company')
plt.ylabel('Jumlah Member')
plt.title('Persebaran Members di Company')

plt.show()
```
