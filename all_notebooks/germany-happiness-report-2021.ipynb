# Focus on Germany

```python
Germany=happy_data2[happy_data2['Country name']=='Germany']
Germany
```

```python
Germany_index= Germany.set_index('year')
Ger = Germany_index.drop(columns='Country name')
Ger.head(20)
```

```python
Ger.info()
```

```python
sns.lineplot(data=Ger['Generosity'], label='Generosity')
plt.title("Development of Generosity in Germany 2006-2020")
plt.xlabel("Year")
plt.ylabel("Score")
```

```python
Ger['Generosity'].fillna(value=Ger['Generosity'].mean(), inplace=True)
sns.lineplot(data=Ger['Generosity'], label='Generosity')
plt.title("Development of Generosity in Germany 2006-2020")
plt.xlabel("Year")
plt.ylabel("Score")
```

```python
sns.lineplot(data=Ger['Social support'], label='Social support')
plt.title("Development of Social support in Germany 2006-2020")
plt.xlabel("Year")
plt.ylabel("Score")
```

```python
sns.lineplot(data=Ger['Log GDP per capita'], label='Log GDP per capita')
plt.title("Development of Log GDP per capita in Germany 2006-2020")
plt.xlabel("Year")
plt.ylabel("Score")
```

```python
sns.lineplot(data=Ger['Healthy life expectancy at birth'], label='Healthy life expectancy at birth')
plt.title("Development of Healthy life expectancy at birth in Germany 2006-2020")
plt.xlabel("Year")
plt.ylabel("Years")
```

```python
sns.lineplot(data=Ger['Freedom to make life choices'], label='Freedom to make life choices')
plt.title("Development of Freedom to make life choices in Germany 2006-2020")
plt.xlabel("Year")
plt.ylabel("Score")
```

```python
sns.lineplot(data=Ger['Perceptions of corruption'], label='Perceptions of corruption')
plt.title("Development of Perceptions of corruption in Germany 2006-2020")
plt.xlabel("Year")
plt.ylabel("Score")
```
