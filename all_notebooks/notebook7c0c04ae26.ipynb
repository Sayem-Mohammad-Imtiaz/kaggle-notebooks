```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
import matplotlib.pyplot as plt
import sklearn

import math
df = pd.read_csv("../input/flight-take-off-data-jfk-airport/M1_final.csv")
df.drop('TAIL_NUM',axis=1,inplace=True)
df.shape
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
df.OP_UNIQUE_CARRIER=l.fit_transform(df.OP_UNIQUE_CARRIER)
df["DEST"]=l.fit_transform(df["DEST"])
df["Wind"] =l.fit_transform(df["Wind"].astype(str))
df["Condition"] =l.fit_transform(df["Condition"])
df["Dew Point"] = l.fit_transform(df["Dew Point"])
#Input and Output
X=df.drop('TAXI_OUT', axis=1)
Y=df.TAXI_OUT
#Splitting
from sklearn.model_selection import train_test_split
X_train,x_test,Y_train,y_test=train_test_split(X,Y,test_size=0.1)
```

```python
# Linear regression(label encoding)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
L_R=LinearRegression()
L_R.fit(X_train,Y_train)
yL_R = L_R.predict(x_test)

L_R_error = mean_squared_error(y_test,yL_R,squared = False)
L_R_error
```

```python
# Ridge regression(label encoding)
from sklearn import linear_model
R_R=linear_model.Ridge(alpha=1,max_iter=1000,tol=.01)
R_R.fit(X_train,Y_train)
yR_R=R_R.predict(x_test)
R_Rerror=mean_squared_error(y_test,yR_R,squared = False)
R_Rerror
```

```python
# Lasso regression(label encoding)
La_R=linear_model.Lasso(alpha=1,max_iter=1000,tol=.01)
La_R.fit(X_train,Y_train)
yLa_R=La_R.predict(x_test)
La_Rerror=mean_squared_error(y_test,yLa_R,squared = False)
La_Rerror
```

```python
# K nearest neighbours(label encoding)
from sklearn.neighbors import KNeighborsRegressor
K_NN = KNeighborsRegressor(n_neighbors = 200)
K_NN.fit(X_train,Y_train)
yK_NN=K_NN.predict(x_test)
K_NNerror = mean_squared_error(y_test,yK_NN,squared = False)
K_NNerror
```

```python
#SVR(label encoding)
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x1=X_train.copy()
x2=x_test.copy()
x1=sc.fit_transform(x1)
x2=sc.fit_transform(x2)
svr=SVR()
svr.fit(x1,Y_train)
ysvr=svr.predict(x2)
svrerror=mean_squared_error(y_test,ysvr,squared = False)
svrerror

```

```python
#Naive Bayes(label encoding)

from sklearn.linear_model import BayesianRidge
N_B=BayesianRidge()
N_B.fit(X_train,Y_train)
yN_B=N_B.predict(x_test)
N_Berror=mean_squared_error(y_test,yN_B,squared=False)
N_Berror
```

```python
#Light GBM(label encoding)

from lightgbm import LGBMRegressor
L_GBM=LGBMRegressor()
L_GBM.fit(X_train,Y_train)
yL_GBM=L_GBM.predict(x_test)
L_GBMerror=mean_squared_error(y_test,yL_GBM,squared=False)
L_GBMerror
```

```python
# Random Forests(label encoding)

from sklearn.ensemble import RandomForestRegressor
R_F = RandomForestRegressor(n_estimators=100)
R_F.fit(X_train,Y_train)
yR_F = R_F.predict(x_test)
R_Ferror = mean_squared_error(y_test,yR_F,squared = False)
R_Ferror
```

```python
# One Hot Encoding
from sklearn.preprocessing import OneHotEncoder
```

```python
p = (df.dtypes == 'object')
cols = list(p[p].index)
OH_e = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_df = pd.DataFrame(OH_e.fit_transform(df[cols]))
OH_cols_df.index = df.index
num_df = df.drop(cols, axis=1)
OH = pd.concat([num_df, OH_cols_df], axis=1)

X=OH.drop(['TAXI_OUT'], axis =1)
Y=OH.TAXI_OUT
from sklearn.model_selection import train_test_split
X_train, x_test, Y_train, y_test = train_test_split( X, Y, test_size=0.1)
```

```python
# Linear regression(One hot encoding)
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
L_Roh=LinearRegression()
L_Roh.fit(X_train,Y_train)
yL_Roh = L_Roh.predict(x_test)
L_Roh_error = mean_squared_error(y_test,yL_Roh,squared = False)
L_Roh_error
```

```python
# Ridge regression(One hot encoding)
from sklearn import linear_model
R_Roh=linear_model.Ridge(alpha=1,max_iter=1000,tol=.01)
R_Roh.fit(X_train,Y_train)
yR_Roh=R_Roh.predict(x_test)
R_Roherror=mean_squared_error(y_test,yR_Roh,squared = False)
R_Roherror
```

```python
# Lasso regression(One hot encoding)
La_Roh=linear_model.Lasso(alpha=1,max_iter=1000,tol=.01)
La_Roh.fit(X_train,Y_train)
yLa_Roh=La_Roh.predict(x_test)
La_Roherror=mean_squared_error(y_test,yLa_Roh,squared = False)
La_Roherror
```

```python
# K nearest neighbours(One hot encoding)
from sklearn.neighbors import KNeighborsRegressor
K_NNoh = KNeighborsRegressor(n_neighbors = 200)
K_NNoh.fit(X_train,Y_train)
yK_NNoh=K_NNoh.predict(x_test)
K_NNoherror = mean_squared_error(y_test,yK_NNoh,squared = False)
K_NNoherror
```

```python
#SVR(One hot encoding)
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x1=X_train.copy()
x2=x_test.copy()
x1=sc.fit_transform(x1)
x2=sc.fit_transform(x2)
svroh=SVR()
svroh.fit(x1,Y_train)
ysvroh=svroh.predict(x2)
svroherror=mean_squared_error(y_test,ysvroh,squared = False)
svroherror
```

```python
#Naive Bayes(One hot encoding)

from sklearn.linear_model import BayesianRidge
N_Boh=BayesianRidge()
N_Boh.fit(X_train,Y_train)
yN_Boh=N_Boh.predict(x_test)
N_Boherror=mean_squared_error(y_test,yN_Boh,squared=False)
N_Boherror
```

```python
#Light GBM (One hot encoding)

from lightgbm import LGBMRegressor
L_GBMoh=LGBMRegressor()
L_GBMoh.fit(X_train,Y_train)
yL_GBMoh=L_GBMoh.predict(x_test)
L_GBMoherror=mean_squared_error(y_test,yL_GBMoh,squared=False)
L_GBMoherror
```

```python
# Random Forests(One hot encoding)

from sklearn.ensemble import RandomForestRegressor
R_Foh = RandomForestRegressor(n_estimators=100)
R_Foh.fit(X_train,Y_train)
yR_Foh = R_Foh.predict(x_test)
R_Foherror = mean_squared_error(y_test,yR_Foh,squared = False)
R_Foherror
```

```python
#plot
from matplotlib import style
E_LE=np.array([L_R_error,R_Rerror,La_Rerror,K_NNerror,svrerror,N_Berror,L_GBMerror,R_Ferror])
E_OHE=np.array([L_Roh_error,R_Roherror,La_Roherror,K_NNoherror,svroherror,N_Berror,L_GBMoherror,R_Foherror])
style.use("ggplot")
plt.plot(["L_R","R_R","La_R","K_NN","SVR","N_B","L_GBM","R_F"],E_LE,"r",label="label encoding",linewidth=2)
plt.plot(["L_R","R_R","La_R","K_NN","SVR","N_B","L_GBM","R_F"],E_OHE,"g",label="one hot encoding",linewidth=2)
plt.title("Comparison")
plt.xlabel("models")
plt.ylabel("root mean squared error")
plt.legend() 
plt.grid(True,color="k")
plt.show
```
