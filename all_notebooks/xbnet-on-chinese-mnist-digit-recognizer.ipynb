```python
import torch
import numpy as np
from sklearn.model_selection import train_test_split
from XBNet.training_utils import training,predict
from XBNet.models import XBNETClassifier
from XBNet.run import run_XBNET 
```

```python
import pandas as pd 
data = pd.read_csv('../input/chinese-mnist-digit-recognizer/chineseMNIST.csv')
data.head()
```

```python
data=data.sample(len(data))
data=data.reset_index(drop = True)
data.head()
```

```python
data.label.value_counts()
```

```python
data.label.unique()
```

```python
data.label=data.label.replace([100,1000,10000,100000000],[11,12,13,14])
data.label.value_counts()
```

```python
y=data[['label']].to_numpy()
y
```

```python
x=data.loc[:,:'pixel_4095'].to_numpy()
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2, random_state= True,stratify=y) 
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
y_train=y_train.reshape((-1))
y_train.shape
```

```python
y_test=y_test.reshape((-1))
y_test.shape
```

```python
model = XBNETClassifier(x_train,y_train,num_layers=5)
```

```python
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.1) 
```

```python
m,acc, lo, val_ac, val_lo = run_XBNET(x_train,x_test,y_train,y_test,model,criterion,optimizer,epochs=32,batch_size=1024)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
plt.subplot(1,2,1)
plt.plot(acc,label='training accuracy')
plt.plot(val_ac,label = 'validation accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()
plt.grid()
plt.subplot(1,2,2)
plt.plot(lo,label='training loss')
plt.plot(val_lo,label = 'validation loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend() 
plt.grid()
```
