```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
```

```python
def checkMetrics(model):
    y_train_pred = model.predict(X_res)
    y_test_pred = model.predict(X_test)
    
    print('Train set:')
    print('Accuracy:', accuracy_score(y_res, y_train_pred))
    print('Recall:', recall_score(y_res, y_train_pred, average = None))
    print('Precision:', precision_score(y_res, y_train_pred, average = None))
    print('F1:', f1_score(y_res, y_train_pred, average = None))
    
    print('Test set:')
    print('Accuracy:', accuracy_score(y_test, y_test_pred))
    print('Recall:', recall_score(y_test, y_test_pred, average = None))
    print('Precision:', precision_score(y_test, y_test_pred, average = None))
    print('F1:', f1_score(y_test, y_test_pred, average = None))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
baseTree = DecisionTreeClassifier(random_state = 42)
baseTree.fit(X_res, y_res)
checkMetrics(baseTree)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {
    'max_depth' : [None, 4, 6, 8, 12, 15, 20, 25, 30],\
    'criterion' : ['entropy', 'gini'],\
    'max_features' : ['auto', 'sqrt', 'log2', None],\
    'max_leaf_nodes' : [None, 2, 4, 6, 8, 10],\
    'min_samples_leaf' : [1, 3, 5, 7, 10, 15, 20],\
    'min_samples_split' : [2, 5, 8, 10, 12, 15, 20]
}
```

```python
clf = GridSearchCV(DecisionTreeClassifier(random_state = 42), params, verbose = 1)
clf.fit(X_res, y_res)
```

```python
clf.best_params_
```

```python
checkMetrics(clf.best_estimator_)
```

```python
tree = DecisionTreeClassifier(random_state = 42, min_samples_leaf = 1, min_samples_split = 20)
tree.fit(X_res, y_res)
```

```python
checkMetrics(tree)
```
