```python
# Read the PriceRunner aggregate file with the 35,000 product titles
df = pd.read_csv('/kaggle/input/product-classification-and-categorization/pricerunner_aggregate.csv', encoding='utf-8', skiprows=1, header=None)
df.columns = ['ID', 'Title', 'VendorID', 'ClusterID', 'ClusterLabel', 'CategoryID', 'CategoryLabel']
df.head(10)
```

```python
# Copy the dataframe to another dataframe with 2 columns
products_df = df[['Title', 'ClusterID']].copy()
products_df.head(10)
```

```python
import re
# Remove punctuation and apply case folding
def preprocessor(text):
    text = re.sub('<[^>]*>', '', text)
    emoticons = re.findall('(?::|;|=)(?:-)?(?:\)|\(|D|P)', text)
    text = (re.sub('[\W]+', ' ', text.lower()) + ' ' .join(emoticons).replace('-', ''))

    return text

# Example on the second row
preprocessor(products_df.loc[1, 'Title'])
```

```python
# apply the preprocessor to all product titles
products_df['Title'] = products_df['Title'].apply(preprocessor)
products_df.head()
```

```python
# we do not want to apply stemming and stop words removal in this dataset
# the following code is just indicative
```

```python
# Copy the Title column of the dataframe into a numpy array
titles_array = products_df['Title'].to_numpy()
titles_array
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(use_idf=True, norm = 'l2', smooth_idf = True)

vectorized_titles = tfidf_vectorizer.fit_transform(titles_array).toarray()
vectorized_titles.shape
```
