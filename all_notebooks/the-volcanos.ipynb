```python
val = df['Name'].value_counts().index
cnt = df['Location'].value_counts().values

fig = go.Figure([go.Bar(x=val, y=cnt, marker_color='darkturquoise')])
fig.update_layout(title_text=' the Volcanos Distribution', title_x=0.5)
fig.show()
```

```python
val = df['Type'].value_counts().index
cnt = df['Location'].value_counts().values

fig = go.Figure([go.Bar(x=val, y=cnt, marker_color='blue')])
fig.update_layout(title_text=' the Volcanos type location', title_x=0.5)
fig.show()
```
