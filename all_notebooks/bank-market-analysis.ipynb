Predictive Analysis of Bank Marketing

**Problem Defination**

The data is related with direct marketing campaigns of a Portuguese banking institution. The marketing campaigns were based on phone calls. Often, more than one contact to the same client was required, in order to access if the product (bank term deposit) would be ('yes') or not ('no') subscribed.

What to achieve?

The classification goal is to predict if the client will subscribe a term deposit (variable y).

Data Contains information in following format:

* Job - (Management,BlueCollar,Technician,entrepreneur,retired,admin.,services,selfemployed,housemaid,student,unemployed,unknown)

* Month - (Jan,Feb,Mar,Apr,May,Jun,Jul,Aug,Sep,Oct,Nov,Dec)
* Poutcome - (Success,Failure,Other,Unknown)
* Housing - (Yes/No)
* Loan - (Yes/No)
* Default - (Yes/No)


**Numerical Variable:**
* Age

Converting object type data into One-Hot Encoded data using get_dummies method.

```python
df_new = pd.get_dummies(df, columns=['job', 'marital', 'education','default',
                                    'housing','loan','contact','month','poutcome'] )
```

```python
df_new.y.replace(('yes','no'),(1,0),inplace = True)
```

```python
#Feature selection

y = pd.DataFrame(df_new['y'])
X = df_new.drop(['y'],axis=1)
```

**Model Selection**

```python
# data divide on training and testing sets

from sklearn.model_selection import train_test_split

X_train, X_test,y_train, y_test = train_test_split(X, y,test_size=0.3, random_state = 2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

**1)Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(X_train, y_train)
```

```python
y_pred = logmodel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

acc_logmodel = round(accuracy_score(y_pred, y_test) * 100)
print(acc_logmodel)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, y_pred))

print('\n')

print(classification_report(y_test, y_pred))
```

**Now we will give a try for Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("Accuracy Score {}".format(accuracy_score(y_test, y_pred)))
```

```python
print("Confusion Matrix \n{}".format(confusion_matrix(y_test, y_pred)))
```

```python
print("Classification Report \n{}".format(classification_report(y_test, y_pred)))
```
