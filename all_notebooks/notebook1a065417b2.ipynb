```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
#graficos
from sklearn.tree import export_graphviz
import graphviz
import matplotlib.pyplot as plt
import numpy as np
```

```python
#importacion del csv
data =pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
#limpio datos nulos
data =data.dropna()
#listados de variables del fichero
data.dtypes
#estadisticos
```

```python
data.describe()
```

```python
print(data)
```

```python
#variables predictoras y objetivas
predictors = data[['Pregnancies','Glucose','BloodPressure','SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']]
targets=data.Outcome
predictors
```

```python
predicions_label=['Pregnancies','Glucose','BloodPressure','SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
taget_label=['1','0']
```

```python
#variable del entrenamiento y variables de prueba
x_entrenamiento, x_test, y_entrenamiento, y_test =train_test_split(predictors,targets)
```

```python
#construcion del arbol
arbol=DecisionTreeClassifier()
arbol.fit(x_entrenamiento, y_entrenamiento)
```

```python
#verificacion de probabilidad de prediccion datos prueba
arbol.score(x_test, y_test)
```

```python
#verificacion de probabilidad de prediccion datos entrenamiento
arbol.score(x_entrenamiento, y_entrenamiento)
```

```python
#generacion de grafica de arbol
export_graphviz(arbol, out_file='arbol.dot', class_names=taget_label, feature_names=predicions_label, impurity=False, filled=True)
with open ('arbol.dot') as f:
    dot_graph=f.read()
graphviz.Source(dot_graph)
```

```python
#importancia de caracteristicas
caract=8
plt.barh(range(caract), arbol.feature_importances_)
plt.yticks(np.arange(caract), predicions_label)
plt.xlabel('importancia de caracteristicas')
plt.ylabel('caracteristicas')
plt.show()
```

```python
#profundidad del arbol actual
arbol.tree_.max_depth
```

```python
#creo nuevo arbol delimitando profundidad(4 segun lo exigido)
arbol=DecisionTreeClassifier(max_depth=4)
```

```python
#entreno nuevamente el arbol
arbol.fit(x_entrenamiento, y_entrenamiento)
```

```python
#nueva presision
arbol.score(x_test, y_test)
```

```python
#generacion de grafica de arbol
export_graphviz(arbol, out_file='arbol.dot', class_names=taget_label, feature_names=predicions_label, impurity=False, filled=True)
with open ('arbol.dot') as f:
    dot_graph=f.read()
graphviz.Source(dot_graph)
```
