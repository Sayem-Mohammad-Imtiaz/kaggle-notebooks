```python
df = pd.read_csv('/kaggle/input/trell-social-media-usage-data/train_age_dataset.csv', index_col='userId')
```

```python
df = df.drop('Unnamed: 0',axis=1)
```

```python
y = df["age_group"]
X = df.drop("age_group", axis=1)
```

```python
X.info()
```

```python
y.unique()
```

```python
y = y - 1
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y)
```

```python
dtrain = lgb.Dataset(X_train, y_train)
dvalid = lgb.Dataset(X_valid, y_valid)
```

```python
params = {'objective': 'multiclass', 'num_class': 4, 'unbalance': True}
model = lgb.train(params, dtrain, num_boost_round=10000, verbose_eval=100, valid_sets=dvalid, early_stopping_rounds=50)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = model.predict(X_valid)
y_pred = y_pred.argmax(axis=1)
```

```python
print(f"f1_score: {f1_score(y_valid, y_pred, average='weighted'):.3f}")
```
