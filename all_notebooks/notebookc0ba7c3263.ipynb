```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error,r2_score,mean_squared_log_error
from sklearn.model_selection import train_test_split
LR=LinearRegression()
RFR=RandomForestRegressor()
DT=DecisionTreeRegressor()
```

```python
df=pd.read_csv('/kaggle/input/eergy-efficiency-dataset/ENB2012_data.csv')
```

```python
df.head()
```

```python
df.columns=['RelativeCompactness','SurfaceArea','Wall Area','Roof Area','Overall Height',
            'Orientation','Glazing Area','Glazing Area Distribution','Heating Load','Cooling Load']
```

```python
x = df.drop(columns=['Heating Load','Cooling Load'])
y1 = df['Heating Load']
y2 = df['Cooling Load']
```

```python
x1_train, x1_test, y1_train, y1_test = train_test_split(x, y1, test_size=0.25, random_state=42)
x2_train, x2_test, y2_train, y2_test = train_test_split(x, y2, test_size=0.25, random_state=42)
```

```python
LR.fit(x1_train,y1_train)
```

```python
y1_pretest=LR.predict(x1_test)
y1_pretest
```

```python
rmse = (mean_squared_error(y1_test,y1_pretest))**0.5
rmse
```

```python
r2_score(y1_test,y1_pretest)
```

```python
RFR.fit(x1_train,y1_train)
```

```python
y1_pretest=RFR.predict(x1_test)
y1_pretest
```

```python
rmse = (mean_squared_error(y1_test,y1_pretest))**0.5
rmse
```

```python
r2_score(y1_test,y1_pretest)
```

```python
DT.fit(x1_train,y1_train)
```

```python
rmse = (mean_squared_error(y1_test,y1_pretest))**0.5
rmse
```

```python
r2_score(y1_test,y1_pretest)
```
