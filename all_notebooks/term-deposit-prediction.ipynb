**Treating Outliers**

```python
l=[0,10,11,12,13]
for i in df.iloc[:,l].columns:
    sns.boxplot(x=df[i],data=df,palette='rainbow')
    
    plt.show()
```

```python
l=[0,10,11]
def Outlier(data):
    for column in data.iloc[:,l].columns:
        if data[column].dtype!='str':
            Q1=np.percentile(data[column],25)
            Q3=np.percentile(data[column],75)
            IQR= Q3-Q1
            lower=Q1-(1.5* IQR)
            upper=Q3+(1.5* IQR)
            data.loc[:,column] =np.where(data[column].values>upper,upper,data[column].values)
            data.loc[:,column] =np.where(data[column].values<lower,lower,data[column].values)
    return data
 
df2=Outlier(df)
df2.columns
df2.head()
```

```python
l=[0,10,11,12,13]
for i in df2.iloc[:,l].columns:
    sns.boxplot(x=df2[i],data=df2,palette='rainbow')
    
    plt.show()
```

**Splitting The Data**

```python
X = df2.drop(['y','contact', 'month', 'day_of_week'] , axis = 1)
y = df2['y']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.30,
                                                    random_state=101)
```

**Building A Model**

```python
#Scaling the data

from sklearn.preprocessing import  MinMaxScaler

mc = MinMaxScaler()

X_train_mc = mc.fit_transform(X_train)

X_test_mc = mc.transform(X_test)
```

```python
from sklearn.svm import SVC

model=SVC()
```

```python
model.fit(X_train_mc,y_train)
```

```python
pred=model.predict(X_test_mc)
pred
```

```python
train_pred=model.predict(X_train_mc)
train_pred
```

**Evaluating The Model**

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
```

```python
accuracy_score(y_test,pred)
```

```python
accuracy_score(y_train,train_pred)
```

```python
print(classification_report(y_test,pred))
```

```python
print(classification_report(y_train,train_pred))
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(confusion_matrix(y_train,train_pred))
```
