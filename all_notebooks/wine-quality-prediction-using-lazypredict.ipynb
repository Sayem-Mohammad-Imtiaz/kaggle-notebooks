```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
import lazypredict
from lazypredict.Supervised import LazyClassifier
```

```python
data = pd.read_csv('../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')

data.head(3)
```

# Simple EDA

```python
data.info(show_counts=True)
```

That's good that we have no missing values.

```python
data.quality.unique()
```

As it was mentioned in dataset discription, I'd like to set quality value which is equal or more than 7 to *1* as good. And if quality of wine is less than 7 we will set to *0* as bad.

```python
data.loc[data.quality >= 7, 'quality'] = 1
# after set 7+ quality to 1, we should also add one extra condition to ignore all 1 values, since there is no original quality = 1 values in the dataset
# it needs to avoid issue when all values in quality column set to 0
data.loc[(data.quality < 7) & (data.quality != 1), 'quality'] = 0
```

```python
plt.title('Quality distribution')
sns.countplot(x='quality', data=data);
```

# Train models

```python
x = data.drop('quality', axis=1)
y = data.quality

x_train, x_test, y_train, y_test = train_test_split(
    x, y,
    test_size=.2,
    random_state=4,
    stratify=y
)
```

```python
model = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models, predictions = model.fit(x_train, x_test, y_train, y_test)
```

# Results

```python
print(models)
```

```python
models['model'] = models.index

models.groupby(['model', 'F1 Score'], as_index=False)['model', 'F1 Score'].mean().sort_values('F1 Score', ascending=False).head()
```

Since **models** variable is dataframe we are able to easily sort results and get **TOP 5** models according to best *F1 Score*. Don't forget that we work with **imbalanced data**!
