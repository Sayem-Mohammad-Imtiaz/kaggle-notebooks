# Import Libraries

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Import traing and testing dataset

```python
dataset_train = pd.read_csv('/kaggle/input/linear-regression-on-logarithm-data/train.csv')
dataset_test = pd.read_csv('/kaggle/input/linear-regression-on-logarithm-data/test.csv')
dataset_train.head()
```

# Extract features and label

```python
X_train = dataset_train.iloc[:, :-1]
y_train = dataset_train.iloc[:, -1]
X_test = dataset_test.iloc[:, :-1]
y_test = dataset_test.iloc[:, -1]
```

# Train simple linear regression model on training dataset

```python
# Import sklearn lib to use the linear regression class
from sklearn.linear_model import LinearRegression

# Initialize the Linear Regression class which can then be used to train the model
regressor = LinearRegression()

# Let's train our Simple Linear Regression model and fit the training dataset to the regressor object
regressor.fit(X_train, y_train)
```

# Predicting on test data

```python
# Let's test our Linear Regression model on test data set
pred = regressor.predict(X_test)
```

# Visualize the training set results

```python
plt.scatter(X_train, y_train, color='green')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title("Training Result (x v/s log(x))")
plt.xlabel("Random Number (x)")
plt.ylabel("Log base 10 value (log(x))")
plt.show()
```

# Visualize the test set results

```python
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, pred, color='blue')
plt.title("Test Result (x v/s log(x))")
plt.xlabel("Random number (x)")
plt.ylabel("Log base 10 value (log(x))")
plt.show()
```
