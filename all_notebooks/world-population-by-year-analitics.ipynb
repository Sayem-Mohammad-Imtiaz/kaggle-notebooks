```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (32,31)

f = open("/kaggle/input/world-population-by-year/WorldPopulation.csv", "r")

tarr = ['2020', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011', '2010', '2009', '2008', '2007', '2006', '2005', '2004', '2003', '2002', '2001', '2000', '1999', '1998', '1997', '1996', '1995', '1994', '1993', '1992', '1991', '1990', '1989', '1988', '1987', '1986', '1985', '1984', '1983', '1982', '1981', '1980', '1979', '1978', '1977', '1976', '1975', '1974', '1973', '1972', '1971', '1970', '1969', '1968', '1967', '1966', '1965', '1964', '1963', '1962', '1961', '1960', '1959', '1958', '1957', '1956', '1955', '1954', '1953', '1952', '1951']
tarr.sort()

population = [0]*len(tarr)
ChangePerc = [0]*len(tarr)
NetChange = [0]*len(tarr)
Density = [0]*len(tarr)
Urban = [0]*len(tarr)

lines = f.readlines()

for i in range(1, len(lines)):
    line = lines[i].replace("\n", "").split(",")
    
    population[tarr.index(line[0])] += int(line[1])-2584034261
    NetChange[tarr.index(line[0])] += int(line[3])-46747398
    Density[tarr.index(line[0])] += int(line[4])-17
    Urban[tarr.index(line[0])] += int(line[5])-775067697

plt.grid(color='grey', linestyle='-', linewidth=0.418)
plt.xlabel("Years")
plt.title("Population")
plt.plot(tarr, population, marker=".", color="#A69A98", label="Population")
plt.plot(tarr, NetChange, marker=".", color="#40B3C0", label="NetChange")
plt.plot(tarr, Density, marker=".", color="#34C436", label="Density")
plt.plot(tarr, Urban, marker=".", color="#C9D419", label="Urban")
plt.legend(prop={"size":18})
plt.show()
```
