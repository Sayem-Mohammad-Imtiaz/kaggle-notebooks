# 2. check artist's song on data

```python
(data['artist']==artist).value_counts()
```

so, we know there is 611 rows of Michael Jackson's work. Now we try to find what is the name of MJ's song.  

```python
df_song_list = data[data['artist']==artist]
df_song_list['song'].value_counts()
```

# 3. Visualize

Now we know there is 41 songs of MJ, and also know value counts.
I want to express the ranking of these songs in a line graph over time.

```python
arr = []
fig = plt.figure(figsize=(40,40))
fig.set_facecolor('white')
ax = fig.add_subplot()
legend = []
for song in df_song_list['song'].unique():
    name_condition = data['song']==song
    artist_condition = data['artist']==artist
    x_value = data[name_condition & artist_condition]['date']
    y_value = data[name_condition & artist_condition]['rank']
    ax.plot(x_value, y_value)
    legend.append(song)
plt.legend(legend)
plt.show()
```
