# **IMPORTING DATASET**

```python
dataset = pd.read_csv("../input/mobile-price-classification/train.csv")
dataset
```

```python
dataset.describe()
```

```python
dataset.info()
```

1. No missing value found.
2. No categorical data found.

```python
sns.heatmap(dataset.corr())
```

Very low correlation between:
1. price_range and mobile_wt
2. price_range and touch_screen
3. price_range and clock_speed

Dropping these columns.

```python
dataset.drop(["mobile_wt", "touch_screen", "clock_speed"], axis = 1)
```

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 1)
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
