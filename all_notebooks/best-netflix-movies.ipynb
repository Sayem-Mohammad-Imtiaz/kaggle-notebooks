```python
#hacemos una tabla de los valores nulos
total_of_all = df.isnull().sum().sort_values(ascending=False)
percent_of_all = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
missing_data_test = pd.concat([total_of_all, percent_of_all], axis=1, keys=['Total Nulos', 'Percent'])
missing_data_test.head()
```

```python
# Movies available in each language

languages_available = pd.DataFrame(df.Language.value_counts())
languages_available
```

```python
#Bar graph of Amount of Movies available in each language (Plotly)

import plotly.express as px

fig = px.bar(languages_available, x=languages_available.index, y='Language')
fig.show()
```

```python
#Highest IMDB Rated Movie In each Language

top_movies = df.groupby(['Language']).apply(lambda df: df.loc[df.IMDB_Score.idxmax()])
top_movies
```

```python
#Highest Movie Rating of each Language Available

plt.figure(figsize=(10,10))
plt.title("Highest Movie Rating of each Language Available")

sns.barplot(x=top_movies['IMDB_Score'], y=top_movies['Language'])
plt.xlabel("IMDB Rating")
plt.ylabel("Language")
```

```python
# TOP Movies Based on IMDB Rating

top_imdb = df.nlargest(5, ['IMDB_Score'])
top_imdb
```

```python
#TOP 5 Movies based on IMDB Rating - Bar Chart

plt.figure(figsize=(10,6))
plt.title("TOP 5 Movies based on IMDB Rating")

sns.barplot(x=top_imdb['IMDB_Score'], y=top_imdb['Title'])
plt.xlabel("IMDB Rating")
plt.ylabel("Movie Title")
```
