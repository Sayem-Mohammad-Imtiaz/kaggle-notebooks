#### This notebook demonstrates a simple data extraction and visualization.

#### We'll extract the "New Deaths" counts for a state, and then compare it to another state.

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
# Make a special time parser to deal with the date format (m/d/Y)
date_parser = lambda x: datetime.strptime(x, "%m/%d/%Y")

# Load the data
df = pd.read_csv('../input/us-covid-deaths-by-state-over-time/United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv', parse_dates=['submission_date'], dayfirst=True, date_parser=date_parser)
```

```python
# Get the data for California
ca = df[df['state'] == "CA"]
ca.head()
```

#### We can see there are lots of columns with specific data. In this case, we're interested in the 'new_death' column.

- Let's plot the CA deaths over the entire pandemic period

```python
# Plot one state's deaths over time
sns.set_theme()
fig, ax = plt.subplots(figsize=(12, 6))
sns.lineplot(data=ca, x="submission_date", y="new_death");
```

#### This gives us a good view of the deaths in CA. Now we'll get Florida as the second state.

```python
# Get the second state
fl = df[df['state'] == "FL"]

# Concat the dataframes together
states = [ca,fl]
state_totals = pd.concat(states)
```

```python
# Plot both state's death counts
new_df = state_totals.reset_index().melt(id_vars=['submission_date', 'state', 'new_death'], var_name='col')

fig, ax = plt.subplots(figsize=(12, 6))
sns.lineplot(x='submission_date', y='new_death', hue='state', palette=['b','r'], data=new_df);
```

#### Now we have a nice comparison between the states.
