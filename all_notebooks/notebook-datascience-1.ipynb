# IMPORTING DATA****

```python
import matplotlib.pyplot as plt
df_orig=pd.read_csv("/kaggle/input/craft-cans/beers.csv")
df=df_orig.copy()
df.describe()


```

```python
df.info()
```

# ****FILLING THE NUN DATA

```python
df=df.fillna(df.mean())



df.describe()
```

# **** Encoding Categorical Data

```python
df["name"].value_counts()
```

```python
df["style"].value_counts()
```

# ****CHECKING THE OUTLIERS

```python
np.where(np.abs(stats.zscore(df["ibu"]))>3)
```

```python
np.where(np.abs(stats.zscore(df["abv"]))>3)
```

```python
np.where(np.abs(stats.zscore(df["ounces"]))>3)
```

# ****INPUTING THE IQR

```python
q1=df.quantile(0.25)
q3=df.quantile(0.75)
IQR=q3-q1
print(IQR)

Top_max=q1-(1.5*IQR)
Bottom_min=q3-(1.5*IQR)
```

# CLEANING THE DATA****

```python
df_clean=df[(df>Bottom_min)|(df<Top_max).any(axis=1)]
```

```python
np.where(np.abs(stats.zscore(df_clean["ibu"]))>3)
```

```python
np.where(np.abs(stats.zscore(df_clean["abv"]))>3)
```

```python
np.where(np.abs(stats.zscore(df_clean["ounces"]))>3)
```
