as we can see here this dataset dosent have any missing value.

now we will seperate categorical variable and numerical variable and then we will find relationship with the target varible price.

```python
categorical_var = [feature for feature in data.columns if data[feature].dtype == 'O' ]
categorical_var
```

```python
numerical_var = [feature for feature in data.columns if data[feature].dtype != 'O' and "year" not in feature]
numerical_var
```

```python
for feature in categorical_var:
    df = data.copy()
    
    df.groupby(feature)["price"].mean().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("Price")
    plt.show()
```

here we clearly see that x7 has highestr price value followed by 8 series , M5 and i8.<br>
also semi-Auto transimission has higher value then automatic and manual car.<br>
further more Hybride fuel type has highets price values somewhere around 26000.

```python
for feature in numerical_var:
    df = data.copy()
    
    df.groupby(feature)["price"].mean().plot.hist()
    plt.xlabel(feature)
    plt.ylabel("Price")
    plt.show()
```
