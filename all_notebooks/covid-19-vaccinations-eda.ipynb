```python
data = dict(type='choropleth', 
            locations = df['ISO3'], 
            z = df['NUMBER_VACCINES_TYPES_USED'], 
            text = df['COUNTRY'],
            colorscale="speed")

layout = dict(title = '<b>Usage of different vaccine types</b>', 
              geo = dict(showframe = False, 
                       projection = {'type':'orthographic'}, 
                       showlakes = True, lakecolor="blue",
                       showocean=True, oceancolor="LightBlue",
                       showland=True, landcolor="darkgray"))


fig = go.Figure(data = data, layout = layout)
fig.update_layout(title_x=0.5, width=700, height=700, font=dict(family="Times New Roman", size=18), font_color="#000000")
po.iplot(fig)
```

```python
vac = []

df["VACCINES_USED"] = df["VACCINES_USED"].astype(str)

for j in df.index:
    lst = df.loc[j,'VACCINES_USED'].split(',')
    for i in lst:
        vac.append(i.strip())
        
import collections

counter = collections.Counter(vac)

most_common = counter.most_common(10)

x = list(dict(most_common).keys())
y = list(dict(most_common).values())

fig = go.Figure(data=[go.Bar(
            x=x, y=y,
            text=y,
            textposition='auto',
            marker_color='rgb(158,202,225)',
            marker_line_color='rgb(8,48,107)',
            marker_line_width=1.5
        )])

fig.update_layout(title= "<b>Most popular vaccine types across the world</b>", title_x=0.5, font=dict(family="Times New Roman", size=18), font_color="#000000")
fig.update_traces(texttemplate='%{text:.2s} countries')

fig.show()
```
