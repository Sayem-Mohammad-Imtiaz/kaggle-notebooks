# Combine Source Files
This script combines the 5 different source files to a single output file.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os

dataframes = []
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        curr_dataframe = pd.read_csv(dirname+'/'+filename)
        dataframes.append(curr_dataframe)

combined_dataframe = dataframes[0]
for i in range(1, len(dataframes)):
    combined_dataframe =  pd.concat([combined_dataframe, dataframes[i]], ignore_index=True, sort=True)

print(combined_dataframe.info())
```
