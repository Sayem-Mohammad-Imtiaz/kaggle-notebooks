# **Importing required libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
```

# **Importing dataset and analysing it**

```python
college = pd.read_csv('../input/forbes-americas-top-colleges-2019/ForbesAmericasTopColleges2019.csv')
```

# **Dropping unwanted columns**

```python
clg = college.drop(columns = ['Name','City','State','Website'])
```

```python
clg.shape
```

# **Checking for null values**

```python
clg.isnull().sum()
```

# **Handling null values with interpolation**

```python
clg['Net Price'] = clg['Net Price'].interpolate() 
clg['Average Grant Aid'] = clg['Average Grant Aid'].interpolate()
clg['Alumni Salary'] = clg['Alumni Salary'].interpolate()
clg['Acceptance Rate'] = clg['Acceptance Rate'].interpolate()
clg['SAT Lower']= clg['SAT Lower'].interpolate()
clg['SAT Upper']=clg['SAT Upper'].interpolate()
clg['ACT Lower']=clg['ACT Lower'].interpolate()
clg['ACT Upper']=clg['ACT Upper'].interpolate()
```

```python
clg.isnull().sum()
```

# **Visualization**

# **Using label binarizer**

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
lb = LabelBinarizer()
```

```python
clgs = lb.fit_transform(clg['Public/Private'])
```

```python
clgs
```

# **Splitting data into Independent variable (x) Dependent variable (y)**

```python
x = clg.iloc[:,2:].values
```

```python
y = clgs
```

# **Splitting data into training and testing sets**

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

# **Scaling x_train and x_test using Standard Scaler**

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_1 = sc.fit_transform(x_train)
```

```python
x_test_1 = sc.fit_transform(x_test)
```

# **Creating activation model**

```python
ann = tf.keras.models.Sequential()
```

# **1st hidden layer**

```python
ann.add(tf.keras.layers.Dense(units = 4, activation = 'relu'))
```

# **2nd hidden layer**

```python
ann.add(tf.keras.layers.Dense(units = 4, activation = 'relu'))
```

# **3rd hidden layer**

```python
ann.add(tf.keras.layers.Dense(units = 4, activation = 'relu'))
```

# **Output layer**

```python
ann.add(tf.keras.layers.Dense(units = 1, activation = 'sigmoid'))
```

# **Defining optimizer and loss function**

```python
ann.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

# **Training our model**

```python
ann.fit(x_train_1, y_train, batch_size = 28, epochs = 65)
```

# **Prediction on testing data**

```python
y_pred = ann.predict(x_test_1)
```

```python
y_pred.shape
```

```python
y_pred = (y_pred > 0.5)
```

```python
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))
```

# **Checking Accuracy**

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
accuracy_score(y_test, y_pred)
```
