```python
# Ltes create our first dataframe
netflix_data = pd.read_csv('../input/netflix-shows/netflix_titles.csv')
netflix_data.head()
```

```python
# Lets take all the years, check what shows or movies were released in each specific year and count them.
n_data = netflix_data.groupby(['release_year', 'type'], as_index=False).show_id.count()
n_data.head()
```

```python
# Lets rename "show_id" to "count", cause that makes more sense right?
n_data.columns = ['release_year', 'type', 'count']
# Here is a tricky part. Coverting our n_data into a pivot table.
# By definition, pivot tables are summary of a more extended table.
n_data_pivot = n_data.pivot(index='release_year', columns='type', values='count').reset_index()
n_data_pivot.fillna(0, inplace = True)  # Filled all the missing values with 0, updated inplace.
n_data_pivot.head()
```

```python
# Declare a chart with one Axis
fig, ax = plt.subplots()  
# Creating some  dummy data
ax.plot(n_data_pivot.release_year, n_data_pivot.Movie, label = 'Movies')
ax.plot(n_data_pivot.release_year, n_data_pivot['TV Show'], label = 'Tv Shows') 
ax.set_ylabel('Number of Movies/TV Shows', fontsize = 14) 
ax.set_xlabel('Year of Release', fontsize = 14)
ax.set_title("Trend Chart", fontsize = 14)
plt.text(x=1970, y=400, s=r'Majority Movies are released after 2000 and Tv Shows mostly after 2010', fontsize = 14)
fig.set_size_inches(18.5, 10.5)
# Lets plot it
plt.grid()
ax.legend(fontsize = 14)
```
