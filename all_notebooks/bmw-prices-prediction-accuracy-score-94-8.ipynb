We'll use get_dummies from pandas to get rid of categorical variables.
But first let's split features and values for prediction ('price').

```python
y=bmw.price
X_=bmw.drop('price',axis=1)
```

```python
X=pd.get_dummies(X_)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =  train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import accuracy_score
```

```python
my_model = XGBRegressor(n_estimators=500, learning_rate=0.1, n_jobs=-1)
```

```python
my_model.fit(X_train, y_train, 
             early_stopping_rounds=5, 
             eval_set=[(X_test, y_test)], 
             verbose=False)
```

```python
y_pred = my_model.predict(X_test)
```

```python
my_model.score(X_test,y_test)
```

So, I've got accuracy score **94.8%** using **XGBRegressor**

```python
prices = pd.DataFrame(data={'Actual':y_test,'Predicted':y_pred}).round(0).astype(int)
```
