### Family section ###


**First of all, We change string data to integer.**

We have 2 string columns , guardian and famsize 


#### Famsize
If family size is GT3 transform '0' or size equal LE3 turn into '1'.As student which live with nuclear family less profit by family attention.


```python
Math_class['Famsize'] = Math_class.famsize.map({'GT3':0,'LE3':1})
```

```python
Por_class['Famsize'] = Math_class.famsize.map({'GT3':0,'LE3':1})
```

#### Guardian

Studen who guardian is mother may have more permission than other so they can do some secret activity with their friend easily.

Becaude of that, We change guardin like that :: 

If student guardian is her\him mother, That's meaning is 1 .

If student guardian is her\him father, That's meaning is 0 .

If guardian is other ,That's meaning is -1.(That data set did not explain relationship between the student and the other. I know that.These statistic is just guess)


```python
Math_class['Guardian'] = Math_class.guardian.map({'father':0,'mother':1,'other':-1})
```

```python
Por_class['Guardian'] = Por_class.guardian.map({'father':0,'mother':1,'other':-1})
```

**Family point**

* So we create a new colums which include family point.

* Family point calcultion :

* Family point = famrel + Famsize + Guardian

```python
def add(val):
    total = 0 
    for i in val:
        total += i 
    return total

Math_class['Family_point'] = Math_class.loc[:,['famrel','Guardian','Famsize']].apply(add,axis=1)
Math_class.loc[:,['Family_point']].head()
```

```python
Por_class['Family_point'] = Por_class.loc[:,['famrel','Guardian','Famsize']].apply(add,axis=1)
Por_class.loc[:,['Family_point']].head()
```

** Total consume ** 

* Lastly, We calculate the total alcohol consume.

* Formula :

* Total alcohol consume = Wacl + Dacl

```python
Math_class['total_C'] = Math_class.loc[:,['Walc','Dalc']].apply(add,axis=1)
Math_class.loc[:,['total_C']].head()
```

```python
Por_class['total_C'] = Por_class.loc[:,['Walc','Dalc']].apply(add,axis=1)
Por_class.loc[:,['total_C']].head()
```

**Graph**

I use the scatter graph because can we understand easily with aplha(transparency).

```python
M_col_1 = Math_class['Family_point']
M_col_2 = Math_class['total_C']

P_col_1 = Math_class['Family_point']
P_col_2 = Math_class['total_C']

fig, ax1 = plt.subplots()
fig, ax2 = plt.subplots()

ax1.scatter(M_col_1,M_col_2,alpha=0.09,color='black')
ax2.scatter(P_col_1,P_col_2,alpha=0.09,color='black')

ax1.set_xlabel('Family point')
ax1.set_ylabel('Alcohol cons')
ax1.set_title('Family\Alcohol(Math)')


ax2.set_xlabel('Family point')
ax2.set_ylabel('Alcohol cons')
ax2.set_title('Family\Alcohol(Port)')
plt.tight_layout()
plt.show()
```

### Relationship section 

In here, We analyze relationship.We separate people according to relotionship and alcohol consume.

Firstly, I use .groupby method in order to find people who has relation ship and what is her/him consumption point.

```python
vote_M = Math_class.groupby(['Walc','romantic'])['sex'].count()
vote_P = Por_class.groupby(['Walc','romantic'])['sex'].count()
```

After that , people was added the list according to relationship. 

```python
M_yes=[]
M_no=[]
P_yes=[]
P_no=[]


for i in range(0,10,2):
    M_no.append(list(vote_M)[i])
    M_yes.append(list(vote_M)[i+1])
    
for i in range(0,10,2):
    P_no.append(list(vote_P)[i])
    P_yes.append(list(vote_P)[i+1])
```

Also, When we used .groupby method , we sort the poeple accotding to point of alcohol consumption. For that we just crate consume list , don't use other method in order to collect data on data set.

```python
cons = [1,2,3,4,5]
```

In the end,We can create bar graph.

```python
x_index = np.arange(len(cons))

plt.bar(x_index,M_yes , width=0.20 , label='yes')
plt.bar(x_index-0.20 , M_no , width=0.20 , label='no')

plt.xticks(ticks=x_index  , labels=cons)

plt.legend(loc='upper right')
plt.title('Consume\Romantic(Math class)')
plt.xlabel('consume')
plt.ylabel('student')
plt.show()

plt.bar(x_index,P_yes , width=0.20 , label='yes')
plt.bar(x_index-0.20 , P_no , width=0.20 , label='no')

plt.xticks(ticks=x_index  , labels=cons)

plt.legend(loc='upper right')
plt.title('Consume\Romantic(Por class)')
plt.xlabel('consume')
plt.ylabel('student')
plt.show()
```
