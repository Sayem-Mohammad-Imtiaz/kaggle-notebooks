```python
def Visualize_first_10(df, catagory):
    # dislay confirmed in first 10 countries 
    df = df.sort_values(catagory, ascending = False)
    fig = px.bar(df.head(10), y = catagory, x = "Country/Region",
            text = catagory, color = "Country/Region")
    #total values bar with 2 precision values
    fig.update_traces(texttemplate = "%{text:.2s}", textposition = "outside")
    # set fontsize and uniformText
    fig.update_layout(uniformtext_minsize = 8)
    fig.update_layout(legend_title_text=f'{catagory} (Covid-19)')
    fig.update_layout(title_text=f'Top 10 {catagory} cases in the world')
    return fig
```

```python
Visualize_first_10(country_wise_df, 'Confirmed')
```

```python
Visualize_first_10(country_wise_df, 'Deaths')
```

```python
Visualize_first_10(country_wise_df, 'Recovered')
```

```python
with plt.style.context('bmh'):
    plt.figure(figsize=(9, 10))
    plt.plot(day_wise_df.Confirmed, '-r', label = 'Confirmed')
    plt.plot(day_wise_df.Deaths, 'b', label = 'Deaths')
    plt.plot(day_wise_df.Recovered, 'y', label = 'Recovered')
    plt.plot(day_wise_df['New cases'], 'c', label = 'New cases')
    plt.plot(day_wise_df['New deaths'], 'k', label = 'New deaths')
    plt.title('Visualized')
    plt.xlabel('Day', fontsize=14)
    plt.ylabel('Cases', fontsize=14)
    plt.legend()
```
