```python
data["Gender"] = data["Gender"].replace("Male",0)
data["Gender"] = data["Gender"].replace("Female",1)
```

```python
data.head()
```

```python
data["MaritalStatus"] = data["MaritalStatus"].replace("Single",0)
data["MaritalStatus"] = data["MaritalStatus"].replace("Partnered",1)
```

```python
data
```

```python
data["Product"] = data["Product"].replace("TM195",0)
data["Product"] = data["Product"].replace("TM498",1)
data["Product"] = data["Product"].replace("TM798",2)
```

```python
data.tail(50)
```

```python
X = data.drop(["Fitness"], axis= 1)
y = data.Fitness
```

```python
print(X.ndim)
print(y.ndim)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train, y_test = train_test_split(X , y, test_size= 0.25, random_state = 200)
```

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
model_1 = LinearRegression()
model_1.fit(X_train,y_train)
```

```python
pred = model_1.predict(X_test)
```

```python
a = []
for i in pred:
    a.append(round(i))
a = np.array(a)
```

```python
import matplotlib.pyplot as plt
print('Coefficients: ', model_1.coef_)
  
# variance score: 1 means perfect prediction
print('Variance score: {}'.format(model_1.score(X_test, y_test)))
  
# plot for residual error
  
## setting plot style
plt.style.use('fivethirtyeight')
  
## plotting residual errors in training data
plt.scatter(model_1.predict(X_train), model_1.predict(X_train) - y_train,
            color = "green", label = 'Train data')
  
## plotting residual errors in test data
plt.scatter(a,a - y_test,
            color = "blue",label = 'Test data')
  
## plotting line for zero residual error
## plotting legend
plt.legend(loc = 'upper right')
  
## plot title
plt.title("Residual errors")
  
## method call for showing the plot
plt.show()
```
