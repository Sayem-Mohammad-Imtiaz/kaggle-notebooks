```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
df=pd.read_csv('/kaggle/input/toronto-subway-delay-data/Toronto-Subway-Delay-Jan-2014-Jun-2021.csv',encoding='utf-8')
```

## Data cleaning

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'])
df['Time']=pd.to_datetime(df['Time'])
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
print("Missing Values:\n")
for col in df.columns:
    missing = df[col].isna().sum()
    percent = missing / df.shape[0] * 100
    print("%s: %.2f%% (%d)" % (col,percent,missing))
```

```python
df['Bound'].value_counts()
```

```python
df['Bound'].fillna('S',inplace=True)
```

```python
df['Line'].value_counts()
```

```python
df['Line'].fillna('YU',inplace=True)
```

```python
df.info()
```

```python
df.dropna(how='any',inplace=True)
```

```python
df.info()
```

```python
#data explore
```

```python
sns.set()
plt.figure(figsize=(12, 6))
sns.pairplot(df)
```

```python
df['Min Delay'].corr(df['Min Gap'])
```

## Conclusion 1: Min Delay and Min Gap are positively correlated

```python
plt.figure(figsize=(20, 6))
df['Day'].value_counts().plot.pie(autopct='%.0f%%')
```

```python
plt.figure(figsize=(20, 6))
df.groupby(df['Day'])['Min Delay'].agg(['mean']).plot.bar()
```

## Conclusion 2: The subway delay has no significant relationship with the day of the week

```python
plt.figure(figsize=(20, 6))
df['Date'].groupby(df['Date']).count().plot()
```

## Conclusion 3: There is a significant change in 2020, The number of delays has obviously decreased

```python
Code_df=df.groupby(df['Code'])['Code'].agg(['count'])
Code_df.shape
```

```python
Code_df=Code_df.sort_values(by=['count'],ascending=False)
Code_df=Code_df.iloc[0:10,:]
Code_df
```

```python
plt.figure(figsize=(20, 6))
plt.bar(Code_df.index,Code_df['count'])
```

## Conclusion 4: Most delay are caused by Code: MUSC, TUSC,MUIS

```python
plt.figure(figsize=(20, 6))
df['Bound'].value_counts().plot.pie(autopct='%.0f%%')
```

## Conclusion 5: Delay mainly happened on bound W

```python
plt.figure(figsize=(20, 6))
df['Line'].value_counts().plot.pie(autopct='%.0f%%')
```
