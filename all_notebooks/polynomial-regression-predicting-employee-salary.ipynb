A Polynomial Regression model for predicting salary of a employee according to their level, more level means higher the position of the employee and hence higher the salary.
Dataset consists of three features,first feature will be ignored because position or level is one of the same thing, hence only level feature will be taken (level is independent and salary is dependent)

```python
#importing libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing dataset
ds = pd.read_csv('../input/position-salary-dataset/Position_Salaries.csv')
ds.head()
```

```python
#since position column and level column are one of the same thing we'll take only level column as our independent variable
x = ds.iloc[:,1:-1].values
y = ds.iloc[:,-1].values
print(x)
print(y)
```

**SPLITTING THE DATASET**
* as only 10 rows are present in the dataset, no splitting will be done so that we can have more data to train our model
* splitting will decrease accuracy

**CREATING A POLYNOMIAL SET WITH DEGREE 4**

```python
from sklearn.preprocessing import PolynomialFeatures
pf = PolynomialFeatures(degree = 4)
x_poly = pf.fit_transform(x)
print(x_poly)
```

```python
#now Linear Regression will be applied on this polynomial set
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_poly,y)
```

**VISUALISING POLYNOMIAL REGRESSION RESULTS**


```python
plt.scatter(x,y,c = 'red')
plt.plot(x,lr.predict(x_poly) )
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.title('Polynomial Regression')
```

```python
#calculating accuracy
from sklearn.metrics import r2_score
r2_score(y,lr.predict(x_poly))
```

**ACCURACY OF THE MODEL CAME 99.7% WHICH IS DUE TO SUCH A SMALL DATASET AND NOT SPLITTING THE DATAEST
