```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (32, 12)
plt.rcParams["font.size"] = 17

f = open("/kaggle/input/pakistan-car-reviews/Pakistan Car Reviews.csv", "r", encoding="ISO8859-5")
labels = []
models = ["Audi", "Nissan", "Suzuki", "KIA", "Honda", "Toyota", "Datsun", "Lexus", "Mitsubishi", "Mercedes", "Hyundai", "FAW", "Chevrolet"]
models += ["Daihatsu", "DFSK", "Isuzu", "BMW", "Chery", "Ford", "ZOTYE", "Changan", "Subaru", "Prince Pearl", "United Bravo", "United Alpha", "Proton"]
models += ["Lamborghini", "Sogo", "Daewoo", "Mazda", "Porsche", "Jeep", "Range Rover", "MG"]

outs = []

for i in range(0, len(models)):
    outs.append([models[i], 0])

clines = []

lines = f.readlines()

for i in range(1, len(lines)):
    line = lines[i].replace("\n", "").split(",")[0]
    if line not in clines:
        clines.append(line)

for i in range(0, len(clines)):  
    
    for el in models:
        if el in clines[i]:
            outs[models.index(el)][1] += 1
            break

outs.sort(key=lambda xx: xx[1], reverse=True)

tarr = []

for i in range(0, len(outs)):
    tarr.append(outs[i][1])
    labels.append(outs[i][0])
    
plt.bar(labels, tarr, color="grey")
plt.title("Number of models.")
plt.grid(color="darkgrey")
plt.xticks(rotation=75)
plt.xlabel("Models")
plt.ylabel("Count model")
plt.show()
```
