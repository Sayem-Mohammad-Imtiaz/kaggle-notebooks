```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
```

```python
data=pd.read_csv("../input/heart-disease-dataset/Heart Disease Dataset.csv");
```

```python
data.describe()
data.info()
```

```python
print(data['target'].unique())
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(), annot = True)
```

```python
accuracies = {}

#X = data.drop('target', axis=1)
X=data.loc[:,('age','sex','cp','exang','oldpeak','slope','ca','thal')].values
Y = data['target']
```

# LABA_2B

# Sigmoid
multylayer perceptron using sigmoid as an activation function: f(x) = 1 / (1 + exp(-x)). 

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)

mlp_logistic = MLPClassifier(hidden_layer_sizes=25, activation='logistic', max_iter = 800)
mlp_logistic.fit(X_train, Y_train)
```

```python
probability_logistic = mlp_logistic.predict_proba(X_test)
print(probability_logistic)
```

```python
Y_pred = mlp_logistic.predict(X_test);
```

```python
mlp_logistic_score = mlp_logistic.score(X_test, Y_test)*100

print('Train accuracy: ','{:.2f}'.format(mlp_logistic.score(X_train, Y_train)*100),"%")
print('Test accuracy: ','{:.2f}'.format( mlp_logistic_score),"%")

accuracies['logistic'] = mlp_logistic_score
```

```python
print(' layers: ',mlp_logistic.n_layers_, '\n','iterations: ',mlp_logistic.n_iter_)
```

# Decision Tree

```python
accuracies3 = {}

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5)

dt_classifier = DecisionTreeClassifier(max_depth = 7, splitter = 'random', min_samples_split = 2)
dt_classifier.fit(X_train, Y_train)

Y_pred = dt_classifier.predict(X_test)
```

```python
dt_classifier_score = dt_classifier.score(X_test, Y_test)*100

print('Train accuracy: ','{:.2f}'.format(dt_classifier.score(X_train, Y_train)*100),"%")
print('Test accuracy: ','{:.2f}'.format( dt_classifier_score),"%")

accuracies3['decision tree'] = dt_classifier_score
```

```python
cm_dt_classifier = confusion_matrix(Y_test,Y_pred)
plt.title("Confusion Matrix")
sns.heatmap(cm_dt_classifier,annot=True,cmap="Greens",fmt="d",cbar=False, annot_kws={"size": 10})
```

# Random Forest

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5)

rf_classifier = RandomForestClassifier(n_estimators = 20, criterion = 'gini')
rf_classifier.fit(X_train, Y_train)

Y_pred = rf_classifier.predict(X_test)
```

```python

rf_classifier_score = rf_classifier.score(X_test, Y_test)*100

print('Train accuracy: ','{:.2f}'.format(rf_classifier.score(X_train, Y_train)*100),"%")
print('Test accuracy: ','{:.2f}'.format( rf_classifier_score),"%")

accuracies3['random forest'] = rf_classifier_score
```

```python
cm_rf_classifier = confusion_matrix(Y_test,Y_pred)
plt.title("Confusion Matrix")
sns.heatmap(cm_rf_classifier,annot=True,cmap="Greens",fmt="d",cbar=False, annot_kws={"size": 10})
```
