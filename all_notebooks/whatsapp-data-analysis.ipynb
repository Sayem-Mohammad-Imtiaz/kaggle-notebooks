```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/whatsapp-chat/Whatsapp_chat.csv')
df.head()
```

```python
df.info()
```

```python
# Changing the column names
df.columns= ['ID','hours', 'days','months','names','timestamp']
```

```python
# dropping unwanted ID column and rearranging columns as needed
df.drop('ID', axis=1)
df=df[['names', 'timestamp','hours','days','months']]
df.head()
```

```python
# checking the unique names
df['names'].unique()
```

```python
df.loc[(df['names'] == 'Person1')].count()['timestamp']
```

```python
# adding num column 
df['num'] = pd.Series([1]*len(df))
```

```python
# counting how many messages sent by each person
df.groupby('names').count()['num']
```

# Which person who was using whatsapp most ?

```python
plt.figure(figsize=(8,6))
sns.countplot(x='names' , data=df) 
plt.show()

# looks like person1 is the most one who was using whatsapp and then person 4
```

# plotting the activation of the people per hour

```python
plt.figure(figsize=(15,6))
plt.title('the activation of the people per hour')
sns.countplot(x='hours',hue= 'names', data=df)
plt.show()

# PERSON 1 : was active during 11pm to 2 am
# PERSON 2 : was active during 1 pm the most
# PERSON 3 : was active during 12 pm to 2am 
# PERSON 4 : was active during 11pm to 2am
```

```python
df['months'].unique()
```

```python
df.groupby('months').count()['num']
```

# plotting the activation per month

```python
plt.figure(figsize=(15,8))
plt.title('The activation of the people per month')
sns.countplot(x='months', hue='names', data=df)
plt.show()

# looks like July  was the most month people were using whatsapp and then june 
```

```python
# add day_of_week column
# first lets convert the timestamp to datetime
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
# and then lets add the column
df['day_of_week'] = df['timestamp'].dt.day_name()
```

```python
df.head()
```

# plotting the activation per day_of_week

```python
plt.figure(figsize=(15,8))
plt.title('the activation per day_of_week')
sns.countplot(x='day_of_week', hue='names', data=df)
plt.show()
# looks like monday is the most activation day and it actually makes sense 
```
