```python
df = pd.read_csv('/kaggle/input/doctor-who-dataset/doctorwho.csv')
simplegraph = nx.from_pandas_edgelist(df,'Source','Target',['Weight'])
```

```python
print("Dw simple graph order: {}\nDw simple graph size: {}".format(simplegraph.order(),simplegraph.size()))
```
