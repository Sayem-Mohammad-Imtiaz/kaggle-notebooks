# **Reading data using Pandas**

```python
df = pd.read_csv('../input/cricinfo-statsguru-data/Men T20I Team Batting Stats.csv')
df
```

# **Splitting the Data to Features (X) and Labels (y)**

```python
X = df.iloc[:,[1,2,3,4,5,7,8]].values
X
```

```python
y = df.iloc[:,6:7].values
y
```

# **Splitting the data into Train and Test Sets**

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25)
```

# **Creating a Deep Learning Model using Tensorflow**

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.BatchNormalization(input_shape=[7,]))
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dense(1))
model.summary()
```

# **Compiling and Training the Deep Learning Model**

```python
model.compile(optimizer='adam',loss='mse')
h = model.fit(X_train,y_train,epochs=500,validation_data=(X_test,y_test))
```

# **Prediction**

```python
p = [[76,18,56,2,0,20.9,7.35]]
pred = model.predict(p)
```
