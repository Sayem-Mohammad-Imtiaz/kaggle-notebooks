# Game of Thrones Exploratory Data Analysis

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../input/game-of-thronesgot/game of thrones.csv" )
```

```python
df.head()
```

```python
plt.figure(figsize=(15,6), dpi=150)
sns.countplot(data = df, x = "Directed by" )
plt.xticks(rotation=90,fontsize=14);
plt.title("Total Episodes By Directors", fontsize =20)
plt.xlabel("Directors", fontsize = 15)
plt.ylabel("No.of Episodes", fontsize =15)
```

```python
plt.figure(figsize=(8,4), dpi=100)
sns.countplot(data = df, x = "Written by", palette="Set1")
plt.xticks(rotation=90,fontsize=8);
plt.title("No. of Episode Written by", fontsize=14);
```

```python
df["Novel(s) adapted"].value_counts()
```

```python
plt.figure(figsize=(14,6), dpi=150)
sns.scatterplot(data=df , x = "Imdb rating", y= "U.S. viewers(millions)",hue="Season",palette="husl", s=100)
plt.legend(title = "Season", loc = (1.05,0.5))
```

### Most Viewed Episode

```python
df.nlargest(1, "U.S. viewers(millions)")
```

### Least Viewed Episode

```python
df.nsmallest(1, "U.S. viewers(millions)")
```

### Most Rated Episode

```python
# Scatter Plot for Season vs Episode
plt.figure(figsize=(14,6), dpi=150)
sns.scatterplot(data=df, x = "Imdb rating", y="Season", hue ="Season",palette = "husl",s=80, legend = False)
plt.title("Season vs Rating Plot", fontsize = 14);
```

```python
# Strip Plot for Season vs Views
fig, ax = plt.subplots(figsize=(14,6),dpi=150)
sns.stripplot(data=df, y = "U.S. viewers(millions)",x ="Season",hue ="Season", ax=ax, s=8)
plt.title("Season vs Views Plot", fontsize = 14);
```
