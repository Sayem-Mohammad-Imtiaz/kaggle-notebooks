```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
train= pd.read_csv('../input/titanic-cleaned-data/train_clean.csv')
test= pd.read_csv('../input/titanic-cleaned-data/test_clean.csv')
titanic= pd.concat([train,test])
#titanic=pd.read_csv('../input/titanic/titanic.csv')
```

```python
titanic
```

```python
titanic.dtypes
```

```python
data=titanic
data['Sex'] = data['Sex'].map({'male':0, 'female':1})
data['Embarked'] = data['Embarked'].fillna('S')
data['Embarked'] = data['Embarked'].map({'S':0, 'C':1, 'Q':2})
data['Family'] = data['SibSp'] + data['Parch']
corr = data.corr()
corr
```

```python
fig, ax = plt.subplots(1, 1, figsize=(7, 7))

mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(corr, square=True,mask=mask,linewidth=2.5, vmax=0.4, vmin=-0.4, cmap='binary', cbar=True, ax=ax)

plt.show()
```
