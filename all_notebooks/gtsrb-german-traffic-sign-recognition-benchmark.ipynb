# **Prepocessing the data**

```python
trainData_demo.ClassId = trainData_demo['ClassId'].astype(str);
```

```python
trainData_demo.ClassId.unique()
```

```python
ImageDataProvider = ImageDataGenerator(rescale=1./225,validation_split=0.2);
```

```python
sampleTrainDaTa = ImageDataProvider.flow_from_dataframe(dataframe=trainData_demo,
directory=inputDir,
x_col="Path",
y_col="ClassId",
subset="training",
batch_size=32,
class_mode='categorical',
target_size=(32, 32))
```

```python
sampleValidDaTa = ImageDataProvider.flow_from_dataframe(dataframe=trainData_demo,
directory=inputDir,
x_col="Path",
y_col="ClassId",
subset="validation",
batch_size=32,
class_mode='categorical',
target_size=(32, 32))
```

```python
image, lable = next(sampleValidDaTa)
```
