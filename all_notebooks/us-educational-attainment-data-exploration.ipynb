```python
# import data visualization library

import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/us-educational-attainment-19952015/1995_2015.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
sns.barplot(x = 'Sex', y = 'Bachelor', data = df, palette = 'coolwarm')
```

Based on the barplot above, and if we disregard those who identify as Both, Females tend to have a slightly higher rate for obtaining a Bachelor degree compared to Males from 1995, 2005, and 2015 collectively.

```python
sns.barplot(x = 'Sex', y = 'Advanced', data = df, palette = 'coolwarm')
```

Interestingly, Males tend to have more Advanced degrees compared to Females.

```python
sns.barplot(x = 'Sex', y = 'Associate', data = df, palette = 'coolwarm')
```

```python
sns.barplot(x = 'Sex', y = 'Some_College_No_Degree', data = df, palette = 'coolwarm')
```

```python
sns.barplot(x = 'Sex', y = 'HS_Diploma', data = df, palette = 'coolwarm')
```

```python
sns.barplot(x = 'Sex', y = 'No_HS_Diploma', data = df, palette = 'coolwarm')
```

Disregarding those who identify as Both, more Males obtain a High School Disploma compared to Females and yet they also have a higher amount when it comes to dropping out of school.

```python
sns.barplot(x = 'Sex', y = 'No_HS_Diploma', data = df, hue = 'Year', palette = 'coolwarm')
```

Again, if we disregard Both, Males have a higher rate for not obtaining a High School Diploma in all three years compared to Females.

```python
sns.barplot(x = 'Sex', y = 'HS_Diploma', data = df, hue = 'Year', palette = 'coolwarm')
```

In 1995, 2005, and 2015, Males also have a higher rate for obtaining a High School Diploma -- this might be due to the fact that there are more Males compared to Females (more data and research would need to be completed in order to confirm this claim).

```python
sns.barplot(x = 'Sex', y = 'Some_College_No_Degree', data = df, hue = 'Year', palette = 'coolwarm')
```

```python
sns.barplot(x = 'Sex', y = 'Associate', data = df, hue = 'Year', palette = 'coolwarm')
```

```python
sns.barplot(x = 'Sex', y = 'Bachelor', data = df, hue = 'Year', palette = 'coolwarm')
```

We saw earlier that collectively, in the years 1995, 2005, and 2015, Females tended to have a slightly higher rate for obtaining their Bachelor degree. However, if we broke the data down into the three different years, Males had a higher rate of obtaining their Bachelor degree in the years 1995 and 2005. It was in the year 2015 that Females outranked Males when it comes to obtaining a Bachelor degree.

```python
sns.barplot(x = 'Sex', y = 'Advanced', data = df, hue = 'Year', palette = 'coolwarm')
```
