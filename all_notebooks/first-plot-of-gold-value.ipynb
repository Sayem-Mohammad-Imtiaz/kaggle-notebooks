```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import math
import seaborn as sns
```

```python
gold = pd.read_csv('../input/gold-prices/gold_v1-5.csv')
commodities = pd.read_csv('../input/gold-prices/commodities - Sheet1.csv')
```

```python
gold['U.S. Official Price\n(U.S. dollars per fine ounce\nend of year)'] = gold['U.S. Official Price\n(U.S. dollars per fine ounce\nend of year)'].replace(np.nan, "$nan")
gold['U.S. Official Price\n(U.S. dollars per fine ounce\nend of year)'] = gold['U.S. Official Price\n(U.S. dollars per fine ounce\nend of year)'].apply(lambda x: float(x[1:]))
```

```python
gold['New York Market Price\n(U.S. dollars per fine ounce)'] = gold['New York Market Price\n(U.S. dollars per fine ounce)'].replace(np.nan, "$nan")
gold['New York Market Price\n(U.S. dollars per fine ounce)'] = gold['New York Market Price\n(U.S. dollars per fine ounce)'].apply(lambda x: x.replace(",", ""))
gold['New York Market Price\n(U.S. dollars per fine ounce)'] = gold['New York Market Price\n(U.S. dollars per fine ounce)'].apply(lambda x: float(x[1:]))
```

```python
gold['London Market Price\n(British £ [1718-1949] or\nU.S. $ [1950-Present] per fine ounce)'] = gold['London Market Price\n(British £ [1718-1949] or\nU.S. $ [1950-Present] per fine ounce)'].replace(np.nan, "$nan")
gold['London Market Price\n(British £ [1718-1949] or\nU.S. $ [1950-Present] per fine ounce)'] = gold['London Market Price\n(British £ [1718-1949] or\nU.S. $ [1950-Present] per fine ounce)'].apply(lambda x: x.replace(",", ""))
gold['London Market Price\n(British £ [1718-1949] or\nU.S. $ [1950-Present] per fine ounce)'] = gold['London Market Price\n(British £ [1718-1949] or\nU.S. $ [1950-Present] per fine ounce)'].apply(lambda x: float(x[1:]))
```

```python
gold = gold.drop(columns = ['U.S. $ [1950-Present] per fine ounce)'])
```

```python
plt.figure(figsize = (14,6))
plt.plot(gold['Year'],gold['U.S. Official Price\n(U.S. dollars per fine ounce\nend of year)'] )
```

```python
plt.figure(figsize = (14,6))
plt.scatter(gold['Year'],gold['New York Market Price\n(U.S. dollars per fine ounce)'] )
plt.scatter(gold['Year'],gold['London Market Price\n(British £ [1718-1949] or\nU.S. $ [1950-Present] per fine ounce)'] )
```
