# Setting Our Envoirnment

```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import sys
import warnings
warnings.filterwarnings("ignore") # For ingnoring warnings
np.set_printoptions(threshold=sys.maxsize)
pd.set_option('display.max_rows', None,'display.max_columns', None,'display.width', None,'display.max_colwidth', -1)
plt.rcParams["figure.figsize"] = (12,5)
```

# Loading Data

```python
df = pd.read_csv("/kaggle/input/pakistanintellectualcapitalcs/Pakistan Intellectual Capital - Computer Science - Ver 1.csv", encoding='ISO-8859-1')
df.head()
```

# Number of Professors per Department

```python
df.Department.groupby(df.Department).count().sort_values(ascending=False).plot(kind="bar")
```

# Degree

```python
df["Terminal Degree"].groupby(df["Terminal Degree"]).count().sort_values(ascending=False)[:30].plot(kind="bar")
```

Most of the professors hold MS, PhD, and BS degrees.

# Specialization

```python
df['Area of Specialization/Research Interests'].groupby(df["Area of Specialization/Research Interests"]).count().sort_values(ascending=False)[:20].plot(kind="bar")
```

Most of them are specialized in Software Engineering.

# Country 

```python
df.Country.groupby(df.Country).count().sort_values(ascending=False)[1:20].plot(kind="bar")
```

Most of the professors are from UK, USA and Malaysia.

# No. of Professors per Province

```python
df["Province University Located"].groupby(df["Province University Located"]).count().sort_values(ascending=False).plot(kind="bar")
```

Most of the professors are teaching in Punjab.
