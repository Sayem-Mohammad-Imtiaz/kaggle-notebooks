```python

le = LabelEncoder()
df_men.iloc[:,3]=le.fit_transform(df_men.iloc[:,3])

Xnew = df_men.iloc[1:264,[10,3]].values
Ynew = df_men.iloc[1:264,4].values 

# Splitting the dataset into the Training set and Test set
X_train, X_test, y_train, y_test = train_test_split(Xnew, Ynew, test_size = 0.2, random_state = 0)
print(df_men.head())
```

```python
sc = StandardScaler()

X_train[:,0: 1] = sc.fit_transform(X_train[:,0:1])

X_test[:,0:1]= sc.transform(X_test[:, 0:1])
```

```python

from sklearn.linear_model import LinearRegression
clf = LinearRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print('linear model coeff (w)', clf.coef_)
print('linear model intercept (b)' , clf.intercept_)
print('R-squared score (training)', clf.score(X_train, y_train))
print('R-squared score (test)', clf.score(X_test,y_test))
```

```python
plt.scatter(X_test[:,0],y_test,color='blue')
plt.scatter(X_test[:,0],y_pred ,color = 'red')
plt.legend(["testing", 'prediction'])
plt.show()


plt.scatter(X_test[:,1],y_test,color='blue')
plt.scatter(X_test[:,1],y_pred ,color = 'red')
plt.legend(["testing", 'prediction'])
plt.show()



```

```python

from sklearn.linear_model import Ridge
linridge = Ridge(alpha=20.0).fit(X_train, y_train)

print('ridge regression linear model intercept: {}'
     .format(linridge.intercept_))
print('ridge regression linear model coeff:\n{}'
     .format(linridge.coef_))
print('R-squared score (training): {:.3f}'
     .format(linridge.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'
     .format(linridge.score(X_test, y_test)))
print('Number of non-zero features: {}'
     .format(np.sum(linridge.coef_ != 0)))
```

```python
yridge=linridge.predict(X_test)
plt.scatter(X_test[:,0],y_test,color='blue')
plt.scatter(X_test[:,0],yridge ,color = 'red')
plt.legend(["testing", 'prediction'])
plt.show()

plt.scatter(X_test[:,1],y_test,color='blue')
plt.scatter(X_test[:,1],yridge ,color = 'red')
plt.legend(["testing", 'prediction'])
plt.show()
```

```python
#Lasso Linear Regression
from sklearn.linear_model import Lasso
linlasso = Lasso(alpha=2.0, max_iter = 10000).fit(X_train, y_train)
print('lasso regression linear model intercept: {}'
     .format(linlasso.intercept_))
print('lasso regression linear model coeff:\n{}'
     .format(linlasso.coef_))
print('Non-zero features: {}'
     .format(np.sum(linlasso.coef_ != 0)))
print('R-squared score (training): {:.3f}'
     .format(linlasso.score(X_train, y_train)))
print('R-squared score (test): {:.3f}\n'
     .format(linlasso.score(X_test, y_test)))
```

```python
y=linlasso.predict(X_test)
plt.scatter(X_test[:,0],y_test,color='blue')
plt.scatter(X_test[:,0],y ,color = 'red')
plt.legend(["testing", 'prediction'])
plt.show()

plt.scatter(X_test[:,1],y_test,color='blue')
plt.scatter(X_test[:,1],y ,color = 'red')
plt.legend(["testing", 'prediction'])
plt.show()


```
