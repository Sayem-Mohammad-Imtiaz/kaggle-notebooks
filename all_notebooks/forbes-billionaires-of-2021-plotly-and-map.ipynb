```python
# Importing required libraries 
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
#Loading dataset
df= pd.read_csv("../input/forbes-billionaires-of-2021/Billionaire.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df['NetWorth'] = df['NetWorth'].str.strip('$')
df['NetWorth'] = df['NetWorth'].str.strip('B')
```

```python
df['NetWorth'] = df['NetWorth'].astype(float)
df['Age'] = df['Age'].astype(np.int64)
```

```python
df1 = df.sort_values(by = ['NetWorth'],ascending = False).reset_index().head(20)

fig = px.bar(df1, x= 'Name', y ='NetWorth', color = 'NetWorth',
             color_continuous_scale = 'sunset',
             hover_data = ['Source','Age'],
             labels= {"NetWorth":"Networth(in Billions)"})

fig.update_layout(title = 'Top 20 Billionaires in the world',
                  title_font = dict(size = 26, family = 'Balto', color = 'Indigo'),
                  title_x = 0.5)

fig.show()
```

```python
df1 = df['Source'].value_counts().reset_index().head(20)

fig = px.bar(df1, x = df1['index'], y = df1['Source'], color =df1['Source'],
             color_continuous_scale ='sunset',
             labels = {"index":"source","Source":"count"})

fig.update_layout(title = 'Top 20 Sources with most Billionaires in the world',
                  title_x = 0.5,
                  title_font = dict(size = 26, family = 'Balto', color = 'Indigo'),
                  xaxis = dict(title = 'Source'),
                  yaxis = dict(title = 'Count')
                 
                 )
fig.show()
```

```python
#This pie shows a breakdown of industries of billionaires around the world

df1 = df['Industry'].value_counts().reset_index()
fig = px.pie(values = df1['Industry'], names = df1['index'], 
             color_discrete_sequence = px.colors.sequential.Sunset)
fig.update_traces(textposition = 'inside', textinfo = 'percent+label')
fig.show()
```

```python
df1 = df['Country'].value_counts().reset_index().head(10)
fig = px.pie(values = df1['Country'], names = df1['index'], 
             color_discrete_sequence = px.colors.sequential.Rainbow)
fig.update_traces(textposition = 'inside', textinfo = 'percent+label')
fig.update_layout(title = 'Top 10 Countries with most number of billionaires',
                  title_x = 0.5,
                  title_font = dict(size = 26, family = 'Balto', color = 'DarkBlue'))
fig.show()
```

```python
df1 = df['Country'].value_counts().reset_index().tail(30)

df1 = df1.rename(columns = {"index":"country"})
df1 = df1.rename(columns = {"Country":"count"})

fig = px.bar(df1, y =  df1['country'], x = df1['count'],color =df1['count'], 
             color_continuous_scale = 'teal',
             orientation = 'h')
fig.update_layout(title = 'Countries with least amount of billionaires',
                  title_x = 0.5,
                  title_font = dict(size = 26, family = 'Balto', color = 'Teal'),
                  xaxis = dict(title = 'Count'),
                  yaxis = dict(title = 'Country'))
fig.show()
```

```python
df1 = df['Country'].value_counts().reset_index()
df1 = df1.rename(columns = {"index":"country"}) 
df1 = df1.rename(columns = {"Country":"count"}) 

fig = px.choropleth(df1, 
                    locations = 'country',
                    locationmode = 'country names',
                    color = 'count',
                    color_continuous_scale = 'twilight')

fig.update_layout(title = 'The number of billionaires in the world',
                  title_x = 0.5,
                  title_font = dict(size = 26, family = 'Balto', color = 'DarkSlateBlue'),
                  geo = dict(showframe = False,
                             showcoastlines = True,
                             projection_type = 'natural earth'))

fig.show()
```

```python
#This map shows the total networth of billionaires in various countries

df1 = df.groupby("Country")['NetWorth'].sum().reset_index()

fig = px.choropleth(df1, 
                    locations = 'Country',
                    locationmode = 'country names',
                    color = 'NetWorth',
                    labels= {"NetWorth":"Networth(in Billions)"},
                    color_continuous_scale = 'orrd')

fig.update_layout(geo = dict(showframe = False,
                             showcoastlines = True,
                             projection_type = 'equirectangular'))

fig.show()
```

```python
#This map shows Top 20 countries by the total networth of billionaires

df1 = df.groupby("Country")['NetWorth'].sum().sort_values(ascending = False).reset_index().head(20)

fig = px.choropleth(df1, 
                    locations = 'Country',
                    locationmode = 'country names',
                    color = 'NetWorth',
                    animation_frame = 'Country',
                    labels= {"NetWorth":"Networth(in Billions)"},
                    color_continuous_scale = 'plasma')

fig.update_layout(geo = dict(showframe = False,
                             showcoastlines = True,
                             projection_type = 'equirectangular'))

fig.show()
```
