### Reading data into a dataframe

````python
df_twlo = pd.read_csv('../input/twilio-stock-price-twlo20162021/TWLO_stock.csv')
````

### Reading stock data
- date -> Date of trading 
- open -> opening price for single stock in USD for the given day
- high -> highest price for single stock in USD for the given day
- close -> closing value for single stock in USD for the given day
- volume -> total stocks (count/NAV)

````python
df_twlo.columns
````

### Renaming the columns for better readibility

````python
df_twlo_renamed = df_twlo.rename(columns = {'1. open':'open', '2. high':'high', '3. low':'low','4. close':'close','5. volume':'volume'}, inplace = False)
````

``` Insight 01 - stock price spiked post pandemic hit (around 2020)```

````python
names = ['open', 'close', 'high', 'low']
fig = plt.figure(figsize=(14,22))

a = 2  # number of rows
b = 4 # number of columns
c = 1  # initialize plot counter

for i in names:
    plt.subplot(a, b, c)
    plt.title('{} displot'.format(i))
    plt.xlabel(i)
    sns.distplot(df_twlo_renamed[i])
    c = c + 1
    
    plt.subplot(a,b,c)
    plt.title('{} boxplot'.format(i))
    plt.xlabel(i)
    plt.boxplot(x=df_twlo_renamed[i])
    c = c + 1
plt.show()
````
