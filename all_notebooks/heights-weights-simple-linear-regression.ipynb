# Separate the Test and Train data sets

```python
dataset = pd.read_csv('/kaggle/input/heights-and-weights/data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
        
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

# Initialize and Train the simple linear regression model

```python
model_candidate = LinearRegression()
model_candidate.fit(X_train, y_train)
```

# Test the Model

```python
y_pred = model_candidate.predict(X_test)
```

# Evaluation & Graphical Analysis

```python
plt.scatter(X_train, y_train, color = "red")
plt.plot(X_train, model_candidate.predict(X_train),  color = "blue")
plt.title("Height vs Weight (Training Data)")
plt.xlabel("Height in meters")
plt.ylabel(" Weight in Kgs")
plt.show()
        
plt.scatter(X_test, y_test, color = "red")
plt.plot(X_train, model_candidate.predict(X_train),  color = "blue")
plt.title("Height vs Weight (Test Data)")
plt.xlabel("Height in meters")
plt.ylabel(" Weight in Kgs")
plt.show()
        
from sklearn.metrics import mean_squared_error
print("Mean Squared Error is" , mean_squared_error(y_test, y_pred))
```
