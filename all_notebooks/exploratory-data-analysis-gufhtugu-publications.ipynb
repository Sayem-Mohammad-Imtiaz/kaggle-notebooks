Renaming column names for better explorations

```python
data = data.rename(columns={'Order Number': 'OrderNumber',"Order Status":"OrderStatus", "Book Name":"BookName","Order Date":"OrderDate","City (Billing)":"BillingCity"})
```

```python
data.head(10)
```

Total number of Books by Gughtugu Publications

```python
totalBooks = data['BookName'].nunique()
print(f'Total Number of Books: {totalBooks}')
```

```python
totalOrderStatus = data.OrderStatus.nunique()
print(f'Total Order Status: {totalOrderStatus}')
```

Number of Books in each Status

```python
data.OrderStatus.value_counts()
```

What is the percentage of Completed orders? Let's check

```python
cPercent = round((data.OrderStatus.value_counts().Completed/totalNumberOfRows)*100,2)
print(f'Percentage of Completed Orders: {cPercent}')

cPercent = round((data.OrderStatus.value_counts().Returned/totalNumberOfRows)*100,2)
print(f'Percentage of Returned Orders: {cPercent}')

clPercent = round((data.OrderStatus.value_counts().Canceled/totalNumberOfRows)*100,2)
print(f'Percentage of Canceled Orders: {clPercent}')
```

Let's see it in a bar chart

```python
data.OrderStatus.explode().value_counts()[:10].plot.bar()
```
