```python
data.columns
```

```python
plt.figure(figsize=(20,6))
sns.barplot(data=data,x='Period',y='Avg rainfall(in mm)')
```

```python
plt.figure(figsize=(20,6))
sns.barplot(data=data,x='Period',y='Avg temp(in centigrade)')
```

avg temp is maximum in july and august ,and in summer as well

```python
plt.plot(data[['Year']],data[['Avg rainfall(in mm)']])
plt.xlabel('year')
```

rain fall is cyclic

```python
fig,ax=plt.subplots(ncols=2,figsize=(20,7))
sns.lineplot(data=data,x='Year',y='Avg rainfall(in mm)',ax=ax[0])
sns.lineplot(data=data,x='Year',y='Avg temp(in centigrade)',ax=ax[1],color='red')
```

```python

sns.lmplot(data=data,y='Avg rainfall(in mm)',x='Avg temp(in centigrade)')
```

we can see that in tempreture range 7.5 -10 centigrade the avg rainfall is maximum(600-1100mm)

```python
data[['Avg rainfall(in mm)','Avg temp(in centigrade)']].corr()
```

there is almost no correlation between temp and rainfall

```python
sns.histplot(data=data,x='Avg rainfall(in mm)')
```

maximum count of rainfall is less than 200
