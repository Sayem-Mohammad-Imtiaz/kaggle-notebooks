```python
import csv
csv_filename = "/kaggle/input/clash-royale-season-18-dec-0320-dataset/BattlesStaging_12272020_WL_tagged/battlesStaging_12272020_WL_tagged.csv"
chunk_size = 500000

text_file_reader = pd.read_csv(csv_filename, engine='python',encoding='utf-8-sig', quoting=csv.QUOTE_MINIMAL, chunksize = chunk_size)
```

```python
dfList = []
counter = 0

for df in text_file_reader:
    dfList.append(df)
    counter= counter +1
    print("Max rows read: " + str(chunk_size * counter) )

df = pd.concat(dfList,sort=False)
```

### Let's Free Up Some Resources

```python
dfList = None
del dfList
```

### Plot the Number Of Matches Per Hour, Y-axis is in LOG scale

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter

df.set_index('battleTime', drop=False, inplace=True)
df.index = pd.to_datetime(df.index)
df_hourly_aggregated = df['arena.id'].groupby(pd.Grouper(freq='60Min')).count()
ax = df_hourly_aggregated.plot(kind='bar', grid=False, figsize=(24,8), color='#86bf91', zorder=2, rot=45)

ax.set_title("Matches Per Hour")
ax.set_xlabel("Date and Hour in UTC", labelpad=20, weight='bold', size=12)
ax.set_ylabel("Matches", labelpad=20, weight='bold', size=12)
ax.set_yscale('log')

# Format y-axis label
ax.yaxis.set_major_formatter(StrMethodFormatter('{x:,g}'))
ax.set_xticklabels([x.strftime("%b %d %Y %H:%M") for x in df_hourly_aggregated.index])
plt.show()
```
