```python
import pandas as pd
import numpy as np
```

```python
import keras 
from keras import layers
from keras.models import Model
from keras.optimizers import SGD

from sklearn.preprocessing import MinMaxScaler
```

```python
#read data 
df=pd.read_csv('../input/housesalesprediction/kc_house_data.csv')
df
```

```python
#Select Important data
df=df[['sqft_lot','floors','price','bathrooms','bedrooms','sqft_living','sqft_above']]
df_x=df[['sqft_lot','floors','bathrooms','bedrooms','sqft_living','sqft_above']]
df_y=df[['price']]

#train_df=df[0:17000]
#test_df=df[17000:]

#preprocessing of data using MinMax
scaler=MinMaxScaler()

scaler.fit(df_x)
new_df_x=scaler.transform(df_x)

scaler.fit(df_y)
new_df_y=scaler.transform(df_y)



```

```python
#Separation of data 
#training data
train_x=new_df_x[0:17000]
train_y=new_df_y[0:17000]

#testing data
test_x=new_df_x[17000:]
test_y=new_df_y[17000:]

print(train_x.shape,train_y.shape)
print(test_x.shape,test_y.shape)
```

```python
#Creation of neural network
inpt=layers.Input(shape=(6,))

x=layers.Dense(100)(inpt)
actv1=layers.Activation('relu')(x)

y=layers.Dense(100)(actv1)
actv2=layers.Activation('relu')(y)

z=layers.Dense(100)(actv2)
actv=layers.Activation('relu')(z)

output=layers.Dense(1)(actv)
model=Model(inpt,output)
```

```python
#plot the model as a graph
keras.utils.plot_model(model, "first_model.png")
```

```python
#display the input and output shapes of each layer in the plotted graph
keras.utils.plot_model(model, "model.png", show_shapes=True)
```

```python
#summary of model
model.summary()
```

```python
#select optimizer
opt=SGD()
```

```python
#compiling the model
model.compile(optimizer=opt,loss='mse')
```

```python
#lunch the learning
model.fit(train_x,train_y,epochs=50,batch_size=64)
```

```python
 #evaluate model
result=model.evaluate(test_x, test_y, batch_size=128)
```

```python
#evaluate model with prediction
predictions=model.predict(test_x)
np.mean((predictions - test_y)**2)
```
