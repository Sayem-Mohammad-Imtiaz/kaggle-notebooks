```python
import pandas as pd
import swachhdata
import swachhdata.text as sdt
```

```python
swachhdata.__version__
```

```python
news = pd.read_csv('/kaggle/input/india-headlines-news-dataset/india-news-headlines.csv')
```

```python
news['headline_text'].head(10)
```

```python
news['headline_text'].tail(10)
```

```python
cr = sdt.ContractionsRecast(verbose=1)
cr.setup(news['headline_text'])
news['headline_text'] = cr.recast()
```

```python
stop = sdt.StopWordsRecast(package='nltk', verbose=1)
news['headline_text'] = stop.setup_recast(news['headline_text'])
```

```python
short = sdt.ShortWordsRecast(min_length=3, verbose=1)
news['headline_text'] = short.setup_recast(news['headline_text'])
```

```python
num = sdt.NumberRecast(process='remove', verbose=1)
news['headline_text'] = num.setup_recast(news['headline_text'])
```

```python
punc = sdt.PunctuationRecast(verbose=1)
news['headline_text'] = punc.setup_recast(news['headline_text'])
```

```python
# StemmingRecast
## method: string (‘porter’, ‘snowball’)

stem = sdt.StemmingRecast(method='snowball', verbose=1)
news['headline_text'] = stem.setup_recast(news['headline_text'])
```

```python
# LemmatizationRecast
## package: string (‘nltk’, ‘spacy’), default=’nltk’

lem = sdt.LemmatizationRecast(package='ntlk', verbose=1)
news['headline_text'] = lem.setup_recast(news['headline_text'])
```
