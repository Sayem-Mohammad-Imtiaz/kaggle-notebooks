# Data Insights and Visualizations

```python
df_new = df.copy()
df_new.head()
```

```python
mushroom_class = {"p": "poisonous", "e": "edible"}
mushroom_cap_shape = {"b":"bell","c":"conical","x":"convex","f":"flat", "k":"knobbed","s":"sunken"}
mushroom_cap_surface = {"f":"fibrous","g":"grooves","y":"scaly","s":"smooth"}
mushroom_cap_color = {"n":"brown","b":"buff","c":"cinnamon","g":"gray","r":"green","p":"pink","u":"purple","e":"red","w":"white","y":"yellow"}
mushroom_bruises = {"t":"bruises","f":"no"}
mushroom_odor = {"a":"almond","l":"anise","c":"creosote","y":"fishy","f":"foul","m":"musty","n":"none","p":"pungent","s":"spicy"}
mushroom_gill_attachment = {"a":"attached","d":"descending","f":"free","n":"notched"}
mushroom_gill_spacing = {"c":"close","w":"crowded","d":"distant"}
mushroom_gill_size = {"b":"broad","n":"narrow"}
mushroom_gill_color = {"k":"black","n":"brown","b":"buff","h":"chocolate","g":"gray", "r":"green","o":"orange","p":"pink","u":"purple","e":"red","w":"white","y":"yellow"}
mushroom_stalk_shape =  {"e":"enlarging","t":"tapering"}
mushroom_stalk_root = {"b":"bulbous","c":"club","u":"cup","e":"equal","z":"rhizomorphs","r":"rooted","?":"missing"}
mushroom_stalk_surface_above_ring = {"f":"fibrous","y":"scaly","k":"silky","s":"smooth"}
mushroom_stalk_surface_below_ring = {"f":"fibrous","y":"scaly","k":"silky","s":"smooth"}
mushroom_stalk_color_above_ring = {"n":"brown","b":"buff","c":"cinnamon","g":"gray","o":"orange","p":"pink","e":"red","w":"white","y":"yellow"}
mushroom_stalk_color_below_ring = {"n":"brown","b":"buff","c":"cinnamon","g":"gray","o":"orange","p":"pink","e":"red","w":"white","y":"yellow"}
mushroom_veil_type = {"p":"partial","u":"universal"}
mushroom_veil_color = {"n":"brown","o":"orange","w":"white","y":"yellow"}
mushroom_ring_number = {"n":"none","o":"one","t":"two"}
mushroom_ring_type = {"c":"cobwebby","e":"evanescent","f":"flaring","l":"large","n":"none","p":"pendant","s":"sheathing","z":"zone"}
mushroom_spore_print_color = {"k":"black","n":"brown","b":"buff","h":"chocolate","r":"green","o":"orange","u":"purple","w":"white","y":"yellow"}
mushroom_population = {"a":"abundant","c":"clustered","n":"numerous","s":"scattered","v":"several","y":"solitary"}
mushroom_habitat = {"g":"grasses","l":"leaves","m":"meadows","p":"paths","u":"urban","w":"waste","d":"woods"}


df_new['class'] = df_new['class'].map(mushroom_class)
df_new['cap-shape'] = df_new['cap-shape'].map(mushroom_cap_shape)
df_new['cap-surface'] = df_new['cap-surface'].map(mushroom_cap_surface)
df_new['cap-color'] = df_new['cap-color'].map(mushroom_cap_color)
df_new['bruises'] = df_new['bruises'].map(mushroom_bruises)
df_new['odor'] = df_new['odor'].map(mushroom_odor)
df_new['gill-attachment'] = df_new['gill-attachment'].map(mushroom_gill_attachment)
df_new['gill-spacing'] = df_new['gill-spacing'].map(mushroom_gill_spacing)
df_new['gill-size'] = df_new['gill-size'].map(mushroom_gill_size)
df_new['gill-color'] = df_new['gill-color'].map(mushroom_gill_color)
df_new['stalk-shape'] = df_new['stalk-shape'].map(mushroom_stalk_shape)
df_new['stalk-root'] = df_new['stalk-root'].map(mushroom_stalk_root)
df_new['stalk-surface-above-ring'] = df_new['stalk-surface-above-ring'].map(mushroom_stalk_surface_above_ring)
df_new['stalk-surface-below-ring'] = df_new['stalk-surface-below-ring'].map(mushroom_stalk_surface_below_ring)
df_new['stalk-color-above-ring'] = df_new['stalk-color-above-ring'].map(mushroom_stalk_color_above_ring)
df_new['stalk-color-below-ring'] = df_new['stalk-color-below-ring'].map(mushroom_stalk_color_below_ring)
df_new['veil-type'] = df_new['veil-type'].map(mushroom_veil_type)
df_new['veil-color'] = df_new['veil-color'].map(mushroom_veil_color)
df_new['ring-number'] = df_new['ring-number'].map(mushroom_ring_number)
df_new['ring-type'] = df_new['ring-type'].map(mushroom_ring_type)
df_new['spore-print-color'] = df_new['spore-print-color'].map(mushroom_spore_print_color)
df_new['population'] = df_new['population'].map(mushroom_population)
df_new['habitat'] = df_new['habitat'].map(mushroom_habitat)
```

# Data Preprocessing

```python
df_2 = df.copy()
df_2.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for c in df_2.columns:
    df_2[c] = le.fit_transform(df_2[c])
```

```python
df_2.head()
```

```python
X = df_2.drop(['class'], axis=1)
y = df_2['class']
print(X.shape, y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=32)
print(X_train.shape, y_train.shape, X_test.shape, y_test.shape)
```

# Model making / Hyperparameter training

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(X,y)
```

```python
feat_importances = pd.Series(model.feature_importances_,index =X.columns)
feat_importances.nlargest(9).plot(kind="barh")
plt.show()
```

Here we can see the importance of all features that are necessary for classification which can also be dedeuced as per the Exploratory data visualization done earlier.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
model_params  = {
    "svm" : {
        "model":SVC(gamma="auto"),
        "params":{
            'C' : [1,10,20],
            'kernel':["rbf"]
        }
    },
    
    "decision_tree":{
        "model": DecisionTreeClassifier(),
        "params":{
            'criterion':["entropy","gini"],
            "max_depth":[5,8,9]
        }
    },
     "random_forest":{
        "model": RandomForestClassifier(),
        "params":{
            "n_estimators":[1,5,10],
            "max_depth":[5,8,9]
        }
    },
    "naive_bayes":{
        "model": GaussianNB(),
        "params":{}
    },
    
    'logistic_regression' : {
        'model' : LogisticRegression(solver='liblinear',multi_class = 'auto'),
        'params': {
            "C" : [1,5,10]
        }
    }
    
}
```

```python
score=[]
for model_name,mp in model_params.items():
    clf = GridSearchCV(mp["model"],mp["params"],cv=8,return_train_score=False)
    clf.fit(X,y)
    score.append({
        "Model" : model_name,
        "Best_Score": clf.best_score_,
        "Best_Params": clf.best_params_
    })
df_scores = pd.DataFrame(score,columns=["Model","Best_Score","Best_Params"])
```

```python
df_scores.iplot(kind='pie', labels='Model', values='Best_Score', title='Model performance', theme='pearl')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
rf = RandomForestClassifier(n_estimators=10, max_depth=8)
rf.fit(X_train, y_train)
```

```python
score = rf.score(X_train, y_train)
score2 = rf.score(X_test, y_test)
score, score2
```

```python
print(classification_report(y_test, rf.predict(X_test)))
```
