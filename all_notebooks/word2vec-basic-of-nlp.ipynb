### We sholud know Word Embedding first, before look at the Word2Vec
### Computers or Machines can't understand Natural Language.
### So, We should change the word to number that they can understand
### That is Word Embedding. Just, Change the word to vecter. (But, we should consider various situations)

```python
# Using Fake and Real news datasets
import pandas as pd
import re
data =pd.read_csv("../input/fake-and-real-news-dataset/True.csv")
data = data['text']

normalized_text = []
for i in data:
    i = i.upper()
    i = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', i)
    normalized_text.append(i.split(' '))

tmp = normalized_text[0]
tmp2=[]
for i in tmp:
    tmp2.append([i])
tmp2[:10]
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder()
pd.DataFrame(enc.fit_transform(tmp2).toarray(),index=tmp)
```

### We ready to make Word2Vec model. So, We will fit it

```python
from gensim.models import Word2Vec
model = Word2Vec(sentences=normalized_text,window=5, min_count=5, workers=4, sg=0)
```

### I will make basic model. So, If you want more imformation. Plz refer [Gensim](https://radimrehurek.com/gensim/models/word2vec.html)

```python
text = model.wv.most_similar('MAN')
text
```

### Input word to model, We can get similar word.

```python
text_tmp = []
for i in text:
    text_tmp.append(i[0])

text_tmp = str(text_tmp)

text_tmp = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ!』\\‘|\(\)\[\]\<\>`\'…》]', '', text_tmp)

from wordcloud import WordCloud
import matplotlib.pyplot as plt

wc1 = WordCloud(max_font_size=200,background_color='white', stopwords='\\',width=800, height=800)
wc1.generate(text_tmp)

plt.figure(figsize=(10, 8))
plt.imshow(wc1)
plt.tight_layout(pad=0)
plt.axis('off')
plt.show()
```
