# **Global map for COVID Vaccinations**

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv', header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()

print(df.columns)

fig = px.choropleth(df,locations='country',locationmode='country names',color='people_fully_vaccinated_per_hundred', hover_name='country', animation_frame='date',color_continuous_scale='rainbow', range_color=(0.0,100.0))
fig.update_layout(title_text="Percentage of the country's total popylation that are fully vaccinated", title_x=0.45)
fig.show()

plotly.offline.plot(fig, filename='vaccination.html')
```

The thing I am visualizing is the percentage of the country's total popylation that are fully vaccinated.

And from the gram we can notice that at aroud December 2020 in German, the vaccine was already getting started. Then slowly lead to the Eurpean area and then North America.

Currently, the most vaccinated country is America with the rate of nearly 40 percentage.

# **Global map of earthquake occurrences worldwild**

```python
df = pd.read_csv('../input/202105-recent-earthquake/query.csv', header=0)
df.index = pd.to_datetime(df['time'])
df['time'] = df.index.strftime('%Y-%m-%d %H:00:00')
df = df.sort_index()

fig = px.scatter_geo(df,lat='latitude',lon='longitude',color='mag', animation_frame='time',color_continuous_scale='rainbow', range_color=(1.0,10.0))
fig.show()
```

We can get the infomation from the picture that, On 24 May, around 14:00, a magnitude 4 earthquake happened in East China Sea, 94 km southwest of Itoman, Okinawa, Japan.

Fortunately, towns or cities near the epicenter where the quake might have been felt as very weak shaking.

And if we notice the occurence of all the earthquakes in this time line, we can breifly see that most earthquakes happened in Pacific Ring of Earthquake. 

In this area, plates collide with each other. In so-called subduction zones, one plate bends and slides under another, causing the ocean's crust to sink into the mantle.

In fact, according to data from the United States Geological Survey (USGS), 81% of the world's strongest earthquakes occur in this active zone.
