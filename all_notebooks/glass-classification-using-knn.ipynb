```python
#Create an empty list called traget
target = []
for i in glass_data['Type']:
    if i >= 1 and i <= 4:
        target.append('1')
    elif i >= 5 and i <= 7:
        target.append('2')
glass_data['Target'] = target
```

```python
glass_data['Target'].unique()
```

```python
print(Counter(glass_data['Target']))
```

```python
glass_data.shape
```

```python
glass_data.head()
#checking up the data set of headers of a file opened
```

```python
glass_data.drop(['Type'],axis=1,inplace =True)
glass_data.shape
```

```python
x= glass_data.iloc[:,:9]
y = glass_data['Target']
```

```python
x.shape
```

# Scale data

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
from sklearn.cross_validation import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25,random_state=2018)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(x_train, y_train)
predict=clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
print(confusion_matrix(y_test,predict))
print(accuracy_score(y_test,predict))
```

# Svm with GridSearchCV

```python
from sklearn import svm
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
```

```python
# Grid Search
# Parameter Grid
param_grid = {'C': [0.1,0.8,0.9,1,1.1,1.2,1.3,1.4], 'gamma': [0.1,0.8,0.9,1,1.1,1.2,1.3,1.4]} 
# Make grid search classifier
clf_grid = GridSearchCV(svm.SVC(), param_grid, verbose=1)
                
# Train the classifier
clf_grid.fit(x_train, y_train)

# clf = grid.best_estimator_()
print("Best Parameters:\n", clf_grid.best_params_)
print("Best Estimators:\n", clf_grid.best_estimator_)
```

```python
model = svm.SVC(kernel='rbf', C = 0.9, gamma=0.1)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
