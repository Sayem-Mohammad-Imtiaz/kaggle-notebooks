```python
#1. Extract your input and output sep
#2. Divide your data into training and testing set
#3. Train your model
#4. Do prediction
```

```python
X=df['YearsExperience'].values.reshape(30,1)
```

```python
y=df['Salary'].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
reg= LinearRegression()
```

```python
reg.fit(X_train,y_train)
```
