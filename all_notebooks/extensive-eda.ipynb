Dataset

- This dataset is originally from the National Institute of Diabetes and Digestive and Kidney Diseases. The objective of the dataset is to diagnostically predict whether or not a patient has diabetes, based on certain diagnostic measurements included in the dataset. Several constraints were placed on the selection of these instances from a larger database. In particular, all patients here are females at least 21 years old of Pima Indian heritage.

- The datasets consists of several medical predictor variables and one target variable, Outcome. Predictor variables includes the number of pregnancies the patient has had, their BMI, insulin level, age, and so on.



```python
data = pd.read_csv('/kaggle/input/pima-indians-diabetes/pima-indians-diabetes.csv', header=None)
data.head()
```

let's understand the columns 

* Pregnancies : Number of times pregnant

* insulin : 2-Hour serum insulin (mu U/ml)
* BMI : Body mass index (weight in kg/(height in m)^2)
* diabetes_pedigree : Diabetes pedigree function
* Age : Age in years (old) 
* Outcome : Class variable (0 or 1) 268 of 768 are 1, the others are 0

```python
# examining the dataset per columns 
# function created to see the dataset, its distribution, std min max etc 
# function will return boxplot and distplot 

import warnings 
warnings.filterwarnings('ignore')
def central_tendency(col):
    mean = data.describe()[col]['mean']
    std = data.describe()[col]['std']
    min_ = data.describe()[col]['min']
    max_ = data.describe()[col]['max']
    median = data.describe()[col]['50%']
    fig, ax = plt.subplots(1, 1, figsize=(16, 7))
    sns.distplot(data[col], bins=20)
    ax.axvline(mean, color='r', linestyle='--', label='mean')
    ax.axvline(median, color='g', linestyle='--', label='medain')
    ax.axvline(max_, color='b', linestyle='-', label='max')
    ax.axvline(min_, color='b', linestyle='--', label='min')

    ax.legend()

    plt.show()
    print(f'Mean of {col} {mean}')
    print(f'std of {col} {std}')
    print(f'min of {col} {min_}')
    print(f'max of {col} {max_}')
    print(f'Median of {col} {median}')
    
    fig, ax = plt.subplots(1, 1, figsize=(16, 7))
    sns.boxplot(data[col],  palette='mako')
    plt.show()
```

### QQ-Plot

#### Theory : 
A Q-Q plot is a scatterplot created by plotting two sets of quantiles against one another. If both sets of quantiles came from the same distribution, we should see the points forming a line that's roughly straight.
