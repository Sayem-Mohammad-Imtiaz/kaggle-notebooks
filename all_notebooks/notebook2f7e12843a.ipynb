```python
import pandas as pd
import numpy as np
import sklearn as sk

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error ,mean_absolute_error, r2_score
```

```python
# Loading data set
df = pd.read_csv('/kaggle/input/diamonds/diamonds.csv')
```

```python
df = df.drop(columns = "Unnamed: 0")
```

```python
df = df.replace("Fair", 1)
df = df.replace("Good", 2)
df = df.replace("Very Good", 3)
df = df.replace("Premium", 4)
df = df.replace("Ideal", 5)

df = df.replace("I1", 1)
df = df.replace("SI2", 2)
df = df.replace("SI1", 3)
df = df.replace("VS2", 4)
df = df.replace("VS1", 5)
df = df.replace("VVS2", 6)
df = df.replace("VVS1", 7)
df = df.replace("IF", 8)

#df = df.drop(columns = "color")
new_color = {'J':1,'I':2, 'H':3,'G':4,'F':5,'E':6,'D':7}
df['color'] = df['color'].map(new_color);
```

```python
df.describe()
```

```python
y = df[["price"]]
X = df.drop(columns = "price")
```

```python
# holding out 30% of data for testing sample
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Linear Regression 
Linear_Reg = LinearRegression()
# fit model 
model = Linear_Reg.fit(X_train, y_train)
y_prediction = Linear_Reg.predict(X_test)

# accuracy score
#print("Model score: ", model.score(y_prediction, y_test) ) 
r2 = r2_score(y_test, y_prediction)
print("r2: ", r2)

# MSE and MSA
mse = mean_squared_error(y_test, y_prediction)
msa = mean_absolute_error(y_test, y_prediction)
print("mse: ", mse)
print("msa: ", msa)
```
