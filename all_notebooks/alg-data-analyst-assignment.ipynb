```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
sns.set_theme(style="ticks", color_codes=True)
```

Loading the dataset

```python
complete_data_set = pd.read_csv("../input/data-analyst-assignment-dataset/Data analyst assignment dataset (make a copy for your analysis) - Member database.csv", dtype={'Year of Graduation':str, 'Graduation Month':str})
```

viewing a small sample size of the data to appreciate the data

```python
complete_data_set.head()
```

checking the number of null values in the data

```python
complete_data_set.isnull().sum()
```

Dropping null values

```python
data = complete_data_set.dropna()
data.head()
```

## A data review

```python
data.describe()
```

```python
data.info()
```

## viewing the gender distribution

```python
data['Gender'].value_counts()
```

```python
sns.catplot(x="Gender", kind="count", palette="ch:.25", data=data)
```

## viewing the institution distribution

```python
data['Institution'].value_counts()
```

## viewing the institution and year of graduation distribution

```python
data.groupby(['Institution','Year of Graduation']).size().unstack().plot(kind='bar',stacked=False)
```

```python
data['Year of Graduation'].value_counts()
```

viewing the trend of number of graduates per year

```python
data['Year of Graduation'].value_counts().plot()
```

## distribution of graduation per year split by gender

```python
data.groupby(['Year of Graduation','Gender']).size()
```

```python
data.groupby(['Year of Graduation','Gender']).size().unstack().plot(kind='bar',stacked=False)
```

## distribution of graduation years

```python
data['Year of Graduation'].value_counts().plot(kind='pie', figsize=(10,10), autopct='%.2f%%')
plt.legend()
plt.show()
```

## distribution of institutions split by gender

```python
data.groupby(['Institution','Gender']).size()
```

```python
data.groupby(['Institution','Gender']).size().unstack().plot(kind='bar',stacked=True)

data['Institution'].value_counts()
```

## distribution of members by institution

```python
sns.catplot(x="Institution", kind="count", palette="ch:.25", data=data)
```

## distribution of members by graduation month

```python
data['Graduation Month'].value_counts().sort_index().plot(kind="bar")
```
