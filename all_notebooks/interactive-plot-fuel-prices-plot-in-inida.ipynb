```python
import numpy as np 
import pandas as pd 
import plotly.offline as pyo
import plotly.graph_objs as go

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('/kaggle/input/fuel-price-in-india/diesel.csv')
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
df['city'].unique()
```

```python
df['state'].unique()
```

```python
df[df['city']=='Mumbai']
```

```python
Mumbai = go.Scatter(
                    x=df[df['city']=='Mumbai']['date'],
                    y=df[df['city']=='Mumbai']['rate'],
                    mode='lines',
                    name='Mumbai')

Delhi = go.Scatter(
                    x=df[df['city']=='Delhi']['date'],
                    y=df[df['city']=='Delhi']['rate'],
                    mode='lines',
                    name='Delhi')

Chennai = go.Scatter(
                    x=df[df['city']=='Chennai']['date'],
                    y=df[df['city']=='Chennai']['rate'],
                    mode='lines',
                    name='Chennai')

Kolkata = go.Scatter(
                    x=df[df['city']=='Kolkata']['date'],
                    y=df[df['city']=='Kolkata']['rate'],
                    mode='lines',
                    name='Kolkata')

Bengaluru = go.Scatter(
                    x=df[df['city']=='Bengaluru']['date'],
                    y=df[df['city']=='Bengaluru']['rate'],
                    mode='lines',
                    name='Bengaluru')

Hyderabad = go.Scatter(
                    x=df[df['city']=='Hyderabad']['date'],
                    y=df[df['city']=='Hyderabad']['rate'],
                    mode='lines',
                    name='Hyderabad')

Coimbatore = go.Scatter(
                    x=df[df['city']=='Coimbatore']['date'],
                    y=df[df['city']=='Coimbatore']['rate'],
                    mode='lines',
                    name='Coimbatore')

data = [Mumbai, Delhi, Chennai, Coimbatore, Hyderabad, Bengaluru, Kolkata]

layout = go.Layout(title='Fuel price',
                  xaxis = dict(title='Year'),
                  yaxis = dict(title='rate'))

fig = go.Figure(data=data, layout=layout)

pyo.iplot(fig)
```
