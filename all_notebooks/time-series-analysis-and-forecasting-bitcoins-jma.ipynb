```python
# importing libreries and changing their name

import itertools
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import matplotlib

plt.style.use('fivethirtyeight')
matplotlib.rcParams['axes.labelsize'] = 14
matplotlib.rcParams['xtick.labelsize'] = 12
matplotlib.rcParams['ytick.labelsize'] = 12
matplotlib.rcParams['text.color'] = 'k'
```

```python
# read the excel file
df = pd.read_csv("../input/bitcoin-trends/CBBTCUSD.csv")
df.head()
```

```python
df.describe()
```

```python
# Sort the Order Date 
df= df.sort_values('DATE')

#print the sorted values
print(df.head(1))

#check any missing values
df.isnull().sum()
```

```python
# grouping sales according to Order Date
df.groupby('DATE')['CBBTCUSD'].sum().reset_index()

# min and max values of Order Date
print(df['DATE'].min())
print(df['DATE'].max())
```

```python
from fbprophet import Prophet
```

```python
df1=df.rename(columns={"DATE": "ds", "CBBTCUSD": "y"})
df1
m = Prophet()
m.fit(df1)
```

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```

```python
from fbprophet.plot import plot_plotly
import plotly.offline as py
py.init_notebook_mode()

fig = plot_plotly(m, forecast)  # This returns a plotly Figure
py.iplot(fig)
```
