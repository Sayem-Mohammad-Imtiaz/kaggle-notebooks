Một cách khác có thể tiếp thị sản phẩm nữa đó là đánh vào nỗi sợ của khách hàng. Hãy cho họ biết họ đang vận động ít như thế nào, đó là một ý tưởng tuyệt hay đấy.

### SLEEP AND TIME IN BED

Chúng ta xem xét mối tương quan giữa lượng thời gian ngủ và thời gian trên giường hằng ngày.

```python
df4
```

```python
sns.relplot(x=df4['TotalMinutesAsleep'],y=df4['TotalTimeInBed'],data=df4)
```

Hầu hết thời gian mọi người nằm trên giường và thời gian ngủ là có mối quan hệ tích cực. Tuy nhiên, có một số phần tử ngoại lai xuất hiện ở đây, lý do hẳn là do họ không thực sự ngủ (hoặc làm việc khác) khi nằm trên giường.

Giúp mọi người theo dõi giấc ngủ và chất lượng giấc ngủ là một ý tưởng hay để tiếp thị tới khách hàng.

#### Gộp 2 bảng dữ liệu với nhau

```python
combined_sleep_day_data=df4.merge(df1,on='Id',how='left')
combined_sleep_day_data.head()
```

```python
combined_sleep_day_data['Id'].nunique()
```

Chỉ có 24 người dùng tích cực sử dụng dữ liệu về giấc ngủ.

```python
combined_sleep_day_data['TotalTimeInBed']
```

#### Thời gian ít vận đồng và thời gian trên giường

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
x=combined_sleep_day_data['TotalTimeInBed'].values.reshape(-1,1)
y=combined_sleep_day_data['SedentaryMinutes'].values.reshape(-1,1)
lm.fit(x,y)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
combined_sleep_day_data['TotalTimeInBed'].corr(combined_sleep_day_data['SedentaryMinutes'])
```

Khi thời gian ở trên giường tăng lên, số phút ít vận động giảm đi, nhưng có vẻ như mối quan hệ này không có ý nghĩa thống kê.
