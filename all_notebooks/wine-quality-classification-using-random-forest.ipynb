**Wine Quality classification using Decision trees and Random forest**


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, confusion_matrix, roc_curve, roc_auc_score
```

```python
df= pd.read_csv('../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
df.head()
```

```python
for i in df.columns:
    sns.distplot(df[i])
    plt.show()
```

```python
df.describe().T
```

```python
df.isnull().sum()
```

```python
df['quality'].value_counts().unique()
```

```python
# Partioning the features into X an Y
x=df.drop(columns = 'quality')
y=df['quality']
x.head()
y.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=541)
```

```python
#Decision trees withoutpreprocessing
clf = DecisionTreeClassifier(max_depth =24, min_samples_leaf= 1, min_samples_split= 2)
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
#Now we try to impute the result taking Entropy
clf2 = DecisionTreeClassifier(criterion = 'entropy', max_depth =24, min_samples_leaf= 1)
clf2.fit(x_train,y_train)
```

```python
clf2.score(x_test,y_test)
```

```python
clf3 = DecisionTreeClassifier(criterion = 'gini', max_depth =24, min_samples_leaf= 1)
clf3.fit(x_train,y_train)
clf3.score(x_test,y_test)
#Entropy calculation gives highest accuracy for this problem.
```

```python
#Now using Random Forest classifier which is also falls under Bagging of Ensemble technique
rand_clf = RandomForestClassifier(random_state=6)
rand_clf.fit(x_train,y_train)
```

```python
rand_clf.score(x_test,y_test)
#Since Random Forest is Ensemble technique taking Decision trees as Base model it able to give a highest accuracy than Decision trees immuning the Outlier impact.
```

```python
#Lets carry out some hyper parametric tuning technique
grid_param = {
    "n_estimators" : [90,100,115,130],
    'criterion': ['gini', 'entropy'],
    'max_depth' : range(2,20,1),
    'min_samples_leaf' : range(1,10,1),
    'min_samples_split': range(2,10,1),
    'max_features' : ['auto','log2']
}
```

So the result shows Random Forest always works better than Decision trees since its an ensemble technique. THe overfitting of Data that is faced in Decision trees is immuned by using Random Forest
