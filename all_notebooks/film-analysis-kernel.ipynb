**We will read data and we will analysis it**

```python
data = pd.read_csv("/kaggle/input/tmdb-movie-metadata/tmdb_5000_movies.csv")
```

**Visualize Data**
* Firstly we will visualize data for understand it.
* I want to compare ratio of revenue and popularity so i put them to graph.
* Then i want to learn correlation of associations so i imported seaborn and i setted area of boxes and i setted up necessary settings

```python
import matplotlib.pyplot as plt
plt.scatter(data.revenue,data.popularity,color = "g", alpha = 0.4)
plt.title("Revenue - Popularity Ratio ")
import seaborn as sns
f,ax = plt.subplots(figsize = (15,15))
sns.heatmap(data.corr(), annot = True, linewidths=5,ax=ax)
plt.show()
```

****Are These films excellent or not bad?**
*  We will evaluate these films and we will decide these films are Excellent or Not Bad. 
* Let's Started
* Firstly - we should get vote_average and we will divide it to length of vote_average of data.
* Second - Finally we know arithmetic average and we can compare films with our average
* Finally - We will associate Evaluation values and original titles
* Yeah it's done.

```python
va = sum(data.vote_average)/len(data.vote_average)
data["Evaluation"] = ["Excellent" if i > va else "Not Bad" for i in data.vote_average]
#print(data["Evaluation"].head(10),data["vote_count"].head(10),data["original_title"].head(10))
Conclusion = list(zip(data["Evaluation"].head(100),data["original_title"].head(100)))
print(Conclusion)
```

```python
Conclusions = [x for x in Conclusion]
Conclusions
```
