```python
import numpy as np
import pandas as pd
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt
```

```python
shipping_data= pd.read_csv('/kaggle/input/ecommerce-data/data.csv')
```

```python
shipping_data.shape
```

```python
shipping_data = shipping_data.drop(labels=range(301, 541909), axis=0)
shipping_data.head
```

```python
data = shipping_data.iloc[:, [3,5]].values
```

```python
dendrogram= sch.dendrogram(sch.linkage(data,'single'))
plt.title('dendrogram')
plt.xlabel('Quantity')
plt.ylabel('UnitPrice')
plt.show()
```

```python
cluster = AgglomerativeClustering(n_clusters=5,affinity='euclidean', linkage='single')
cluster.fit_predict(data)
```

```python
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
plt.title('Clusters shipments')
plt.xlabel('Quantity')
plt.ylabel('UnitPrice')
```
