# Importing Libraries

```python
#Data Analysis
import pandas as pd
import numpy as np
#Data Visualisations
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Read the data
data=pd.read_csv('../input/superstore-sales-dataset/SuperStore_Orders.csv')
data.head(5)
```

# Exploratory Data Analysis

```python
#Checking for null values and stats
data.info()
```

```python
data.describe()
```

#Extracting Data Only for US

```python
data_main= data[data['market']=='US']
data_main.shape
```

```python
#Dropping columns that are not required (dropping country column since the data is only for US)
data_main=data_main.drop('country',axis=1)
data_main.shape
```

```python
#Top 10 products in each category
```

```python
#converting sales to float datatype
data_main['sales']=data_main['sales'].str.replace(',','')
data_main['sales']=data_main['sales'].astype(float)
data_main['sales'].sum()
```

```python
#Top category by quantity
sns.countplot(data=data_main,x='category',palette='dark')
```

```python
top_products = pd.DataFrame(data_main.groupby('category').sum()['sales'])
#Top category by currency sales
top_products.plot(kind='bar',color='#b80045')
```

```python
#Products in tech category are sold the least however they drive the highest currency sales
```

```python
plt.figure(figsize=(16,8))
plt.bar('sub_category','category',data=data_main,color='y')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.pie(data_main['sub_category'].value_counts(),autopct="%1.1f%%")
plt.show()
```

```python
#Profit vs Sales
value = data_main.groupby('sub_category').sum()[['sales','profit']]
value.plot(kind='bar',figsize=(16,5),color=['m','b'])
```

```python
#Chairs and Phones have the highest sales while the highest profit is driven from copiers!
```

```python
#Purchasing trends for each region
```

```python
plt.figure(figsize=(16,5))
sns.countplot(x='sub_category',hue='region',data=data_main)
```
