```python
test_data = pd.read_csv('../input/titanic-competition/test.csv')
test_data.head()
```

```python
test_data.loc[((test_data['SibSp'] + test_data['Parch']) == 0 ), 'Alone'] = 'yes'  
test_data.loc[((test_data['SibSp'] + test_data['Parch']) != 0 ), 'Alone'] = 'no' 
```

```python
test_data['Cabin_#']= test_data['Cabin'].str.get(0)
```

```python
test_data.loc[test_data['Age'] <=16, 'person'] = 'children' 
test_data.loc[test_data['Age'].isnull(), 'person'] = test_data['Sex'] 
test_data.loc[test_data['Age']>16, 'person'] = test_data['Sex'] 
```

```python
test_data.loc[test_data['Pclass'] ==1, 'Pclass_P'] =p1_prob
test_data.loc[test_data['Pclass'] ==2, 'Pclass_P'] =p2_prob
test_data.loc[test_data['Pclass'] ==3, 'Pclass_P'] =p3_prob
test_data.loc[test_data['Pclass'].isnull(), 'Pclass_P'] =total_prob
```

```python
test_data.loc[test_data['person'] =='male', 'Person_P'] =m_prob
test_data.loc[test_data['person'] =='female', 'Person_P'] =f_prob
test_data.loc[test_data['person'] =='children', 'Person_P'] =c_prob
#test_data.loc[test_data['person'].isnull(), 'Pclass_P'] =total_prob
```

```python
def age_p(slope, intercept, x):
    return (x*slope + intercept)*total_prob/A_prob_not_null
```

```python
test_data.loc[test_data['Age'].notnull(), 'Age_P'] =age_p(slope, intercept,test_data['Age'] )
test_data.loc[test_data['Age'].isnull(), 'Age_P'] =total_prob
```

```python
test_data.loc[test_data['Alone'] =='yes', 'Alone_P'] =a_prob
test_data.loc[test_data['Alone'] =='no', 'Alone_P'] =na_prob
```

```python
test_data.loc[test_data['Embarked'] =='C', 'Embarked_P'] =ec_prob*total_prob/enn_prob
test_data.loc[test_data['Embarked'] =='Q', 'Embarked_P'] =eq_prob*total_prob/enn_prob
test_data.loc[test_data['Embarked'] =='S', 'Embarked_P'] =es_prob*total_prob/enn_prob
```

```python
test_data['Cabin_loc']= test_data['Cabin'].str.get(0)
test_data.loc[test_data['Cabin_loc'] =='A', 'Cabin_P'] =Ca_prob*total_prob/Cnn_prob
test_data.loc[test_data['Cabin_loc'] =='B', 'Cabin_P'] =Cb_prob*total_prob/Cnn_prob
test_data.loc[test_data['Cabin_loc'] =='C', 'Cabin_P'] =Cc_prob*total_prob/Cnn_prob
test_data.loc[test_data['Cabin_loc'] =='D', 'Cabin_P'] =Cd_prob*total_prob/Cnn_prob
test_data.loc[test_data['Cabin_loc'] =='E', 'Cabin_P'] =Ce_prob*total_prob/Cnn_prob
test_data.loc[test_data['Cabin_loc'] =='F', 'Cabin_P'] =Cf_prob*total_prob/Cnn_prob
test_data.loc[test_data['Cabin_loc'] =='G', 'Cabin_P'] =Cg_prob*total_prob/Cnn_prob
test_data.loc[test_data['Cabin_loc'].isnull(), 'Cabin_P'] =total_prob
```

```python
total_prob/Cnn_prob
```

```python
test_data['P_combined']=(test_data['Pclass_P']+test_data['Person_P']+test_data['Age_P']+test_data['Alone_P']+test_data['Embarked_P']+test_data['Cabin_P'])/6
```

```python
test_data['P_combined'].mean()
```

```python
test_data.loc[test_data['P_combined'] >total_prob, 'Survived'] = int(1)
test_data.loc[test_data['P_combined'] <=total_prob, 'Survived'] = int(0)

```

```python
test_data = test_data[['PassengerId', 'Survived_Assumption']]
test_data.rename(columns={'Survived_Assumption': 'Survived'}, inplace=True)
```
