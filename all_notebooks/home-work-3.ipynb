```python
df = pd.read_csv('/kaggle/input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
df.head()
```

Target - оценить качество вина (признак - quality)
Все признаки количественные
Масштабирование для деревьев не требуется

```python
y = df['quality'] 
X = df.drop('quality', axis=1)

X.head()
```

Разбиваем датасет на тренировачную и обучающую выборки

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X, y, 
                                                      test_size=0.3, random_state=2019)
```

Обучим нашу модель 

```python
from sklearn.tree import DecisionTreeClassifier

tree = DecisionTreeClassifier(max_depth=5, random_state=2019)
tree.fit(X_train, y_train)
```

Оценим качетсво модели

```python
from sklearn.metrics import accuracy_score

y_pred = tree.predict(X_valid)
accuracy_score(y_valid, y_pred)
```
