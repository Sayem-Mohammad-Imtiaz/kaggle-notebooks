```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Activation
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import to_categorical
from keras.preprocessing import image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from tqdm import tqdm
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```python
data = pd.read_csv("/kaggle/input/age-gender-and-ethnicity-face-data-csv/age_gender.csv")
```

```python
data.head()
```

```python
data['pixels']=data['pixels'].apply(lambda x:  np.array(x.split(), dtype="float32"))
```

```python
data.head()
```

```python
data["gender"].value_counts()
```

```python
#Here X = np.array(data['pixels'].tolist()) .tolist() coz it is already in a nd array so we are converting to list and then again converting this single column as array inorder to reshape it.
```

```python
#numpy.reshape(a, newshape, order='C') here a => array to be reshaped here X has only 1column so X[0].
```

```python
X = np.array(data['pixels'].tolist())

## Converting pixels from 1D to 3D
X = X.reshape(X.shape[0],48,48,1)
```

```python
y = data["gender"]
```

```python
y = to_categorical(y)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = Sequential()
```

```python
input_shape = (X.shape[1],X.shape[2],X.shape[3])
```

```python
input_shape
```

```python
model.add(Conv2D(input_shape = (X.shape[1],X.shape[2],X.shape[3]),filters=3,kernel_size=(3,3),strides=(1,1),padding="same"))
```

```python
model.output_shape
```

```python
import tensorflow as tf
```

```python
#Activation Layer
model.add(Activation("relu"))
#Pooling Layer
model.add(MaxPooling2D(pool_size=(2,2)))
```

```python
model.add(Flatten())
```

```python
model.output_shape
```

```python
model.add(Dense(50))
model.add(Activation('relu'))
model.add(Dense(50))
model.add(Activation('relu'))
model.add(Dense(50))
model.add(Activation('relu'))
model.add(Dense(50))
model.add(Activation('relu'))
model.add(Dense(2))
model.add(Activation('softmax'))
```

```python
from keras import optimizers
```

```python
adam = optimizers.Adam(lr=0.001)
model.compile(loss="categorical_crossentropy",optimizer=adam,metrics=["accuracy"])
```

```python
history = model.fit(X_train,y_train,batch_size=50,epochs=50,validation_split=0.3,verbose=1)
```

```python
results = model.evaluate(X_test,y_test)
results[1]
```
