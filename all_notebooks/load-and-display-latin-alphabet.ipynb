```python
import numpy as np # linear algebra
import zipfile
import matplotlib.pyplot as plt
import string

# Load images and labels
X = np.loadtxt('/kaggle/input/comnist/latin_data/latin_data.csv', delimiter=",")
y = np.loadtxt('/kaggle/input/comnist/latin_label/latin_label.csv', delimiter=",")
```

```python
# Display 5 random letters and their labels
for _ in range(5):
    i = np.random.randint(X.shape[0])
    my_letter = X[i].reshape(28,28)
    my_label = string.ascii_uppercase[int(y[i])]
    print(f"Displaying letter {my_label}")
    plt.imshow(my_letter, cmap="gray")
    plt.show()
```
