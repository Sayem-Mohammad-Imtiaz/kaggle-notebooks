```python
# This Python 3 environment with analytics libraries installed
# as defined by the kaggle/python Docker 
import numpy as np 
import pandas as pd 

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#visualization libraries
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
df=pd.read_csv('../input/cyber-incidents-up-to-2020/cyber-operations-incidents.csv')
```

```python
df["Date"] = pd.to_datetime(df["Date"])
#focus on the 2020 incidents
#remove those rows where Date in NaN
df2020 = df[pd.notnull(df['Date'])]
```

```python
#select dates in the year 2020
start_date = '2020-01-01'
end_date = '2020-12-31'

yr2020 = (df2020['Date'] > start_date) & (df2020['Date'] <= end_date)
cyber2020 = df2020.loc[yr2020]

#sorted by date
cyber2020 = cyber2020.sort_values(by='Date')
#Reset row indexes of the DataFrame
#to ignore those rows indexes previously removed

cyber2020 = cyber2020.reset_index(drop=True)
```

```python
cyber2020=cyber2020.drop(['Sources_1', 'Sources_2', 'Sources_3'], axis = 1) 
```

```python
US1 = cyber2020[(cyber2020['Victims'] == 'The U.S. electrical grid')] 
US2 = cyber2020[(cyber2020['Victims'] == 'Employees of the U.S. government')] 
US3 = cyber2020[(cyber2020['Victims'] == 'U.S. government employees and possibly Westat employees')]
US4 = cyber2020[(cyber2020['Victims'] == 'U.S. government agencies, U.S. companies')] 
US5 = cyber2020[(cyber2020['Victims'] == 'U.S. defense contractors')] 
US6 = cyber2020[(cyber2020['Victims'] == 'Individuals associated with the U.S. government or upcoming election')]
US7 = cyber2020[(cyber2020['Victims'] == 'Individuals associated with the U.S. election')] 
US8 = cyber2020[(cyber2020['Victims'] == 'Organizations and individuals associated with U.S. politics')] 
US9 = cyber2020[(cyber2020['Victims'] == 'U.S. government and private companies')] 
US10 = cyber2020[(cyber2020['Victims'] == 'U.S. government')] 
US11 = cyber2020[(cyber2020['Victims'] == 'U.S. government networks')] 
```

```python
US = [US1,US2, US3, US4, US5, US6, US7, US8, US9, US10, US11]
USincidents= pd.concat(US)
USincidents=USincidents.reset_index(drop=True)
USincidents
```
