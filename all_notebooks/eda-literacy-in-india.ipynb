<h2 style='text-align:center;font-size:30px;background-color:#FE7176;border:20px;color:white'>Univariate Analysis<h2>

```python
fig = go.Figure(data=go.Violin(y=df['Literacy'], box_visible=True, line_color='black',
                               meanline_visible=True, fillcolor='lightseagreen', opacity=0.6,
                               x0='Literacy'))

fig.update_layout(yaxis_zeroline=False)
fig.show()
```
