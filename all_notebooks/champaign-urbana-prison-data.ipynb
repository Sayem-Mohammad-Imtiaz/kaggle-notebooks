```python
#Update the release reason to reflect it's missing
df['release_reason'] = df['release_reason'].fillna("Missing Data")
df['release_reason_description'] = df['release_reason_description'].fillna("Missing Data")
df['inmate_status'] = df['inmate_status']
```

Since it's unclear if the missing data is left blank intentionally, accidentally, or actually means N/A, we are going to work around it for now, rather than filling it in when we don't know the categorical data.

```python
# distribution of how long the average arrest is in ja
df['days_in_jail'].hist()
```

```python
# See how many didn't spend a night in jail
zero_count = df[df.days_in_jail == 0]
zero_count
print(len(zero_count))
```

```python
# See how many did 1-3 nights
one_count = df[df.days_in_jail == 1]
print(len(one_count))
two_count = df[df.days_in_jail == 2]
print(len(two_count))
three_count = df[df.days_in_jail == 3]
print(len(three_count))
```

Of the ~68,000 items in the data set, about 40% (26,210) weren't in jail overnight, and about 25% (16,079) spent one night in jail.
