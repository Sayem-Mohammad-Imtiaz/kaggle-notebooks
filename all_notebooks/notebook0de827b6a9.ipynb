```python
import pandas as pd
```

```python
df=pd.read_csv('/kaggle/input/car-bay-data-set/Car_Buy.csv')
df.head()
```

```python
inputs = df.drop('Buy',axis='columns')
```

```python
target = df['Buy']
```

```python
from sklearn.preprocessing import LabelEncoder
le_Age = LabelEncoder()
le_Mil = LabelEncoder()
le_RoadTested = LabelEncoder()
```

```python
inputs['Age_n'] = le_Age.fit_transform(inputs['Age'])
inputs['Mileage_n'] = le_Mil.fit_transform(inputs['Mileage'])
inputs['RoadTested_n'] = le_RoadTested.fit_transform(inputs['Road tested'])
```

```python
inputs_n = inputs.drop(['Age','Mileage','Road tested'],axis='columns')
```

```python
inputs_n
```

import sklearn and make your model 


```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
#model = tree.DecisionTreeClassifier(criterion='entropy')
```

```python
model.fit(inputs_n, target)
```

```python
model.score(inputs_n,target)
```

Sould I buy recent car with Low mileage which is not road tested?

```python
model.predict([[1,1,0]])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_n, target, test_size=0.20, random_state=101)
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
print(X_test)
print(y_test)
print(predictions)
```

```python
model.score(X_test,y_test)
```
