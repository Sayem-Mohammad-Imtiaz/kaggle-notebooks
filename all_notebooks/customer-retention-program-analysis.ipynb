```python
# data=pd.read_csv(r"WA_Fn-UseC_-Telco-Customer-Churn.csv")
d=data.drop(['customerID','Partner','TotalCharges'],axis=1)
d.dropna(inplace=True)
# d['gender'].apply(str)
d['gender']=data['gender'].apply(lambda x: '1' if x=='Male' else '0'  )
d['Dependents']=data['Dependents'].apply(lambda x: '1' if x=='Yes' else '0'  )
# d['SeniorCitizen']=data['SeniorCitizen'].apply(lambda x: '1' if x=='No' else '0'  )
d['tenure']=data['tenure'].apply(lambda x: '1' if x=='0-16' else '2'if x=='16-34' else '3'if x=='34-64' else '0' )
d['PhoneService']=data['PhoneService'].apply(lambda x: '1' if x=='Yes' else '0'  )
d['MultipleLines']=data['MultipleLines'].apply(lambda x: '1' if x=='Yes' else '2' if x== 'No internet service' else '0')
d['InternetService']=data['InternetService'].apply(lambda x: '1' if x=='Fiber optic' else '2' if x =='DSL' else'0' )
d['OnlineSecurity']=data['OnlineSecurity'].apply(lambda x: '1' if x=='Yes' else '2' if x== 'No internet service' else '0')
d['OnlineBackup']=data['OnlineBackup'].apply(lambda x: '1' if x=='Yes' else '2' if x== 'No internet service' else '0')
d['DeviceProtection']=data['DeviceProtection'].apply(lambda x: '1' if x=='Yes' else '2' if x== 'No internet service' else '0')
d['TechSupport']=data['TechSupport'].apply(lambda x: '1' if x=='Yes' else '2' if x== 'No internet service' else '0')
d['StreamingTV']=data['StreamingTV'].apply(lambda x: '1' if x=='Yes' else '2' if x== 'No internet service' else '0')
d['Contract']=data['Contract'].apply(lambda x: '1' if x=='Month-to-month' else '2' if x== 'Two year' else '0')
d['MonthlyCharges']=data['MonthlyCharges'].apply(lambda x: '1' if x=='64-118' else '2'if x=='34-64' else '0' )
d['Churn']=data['Churn'].apply(lambda x: '1' if x=='Yes' else '0'  )
d['StreamingMovies']=data['StreamingMovies'].apply(lambda x: '1' if x=='Yes' else '2' if x== 'No internet service' else '0')
d['PaperlessBilling']=data['PaperlessBilling'].apply(lambda x: '1' if x=='Yes' else '2' if x== 'No internet service' else '0')
d['PaymentMethod']=data['PaymentMethod'].apply(lambda x: '1' if x=='Bank transfer (automatic)' else '2' if x== 'Electronic check' else '0')
X_Churn=d.drop(['Churn'],axis=1)
Y_Churn=d['Churn']
d.head(10)
```
