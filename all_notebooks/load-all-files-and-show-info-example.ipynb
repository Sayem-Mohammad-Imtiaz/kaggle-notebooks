```python
import os
# for dirname, _, filenames in os.walk('/kaggle/input'):
#     for filename in filenames:
#         print(os.path.join(dirname, filename))
path_loc = '/kaggle/input/scrnaseq-bone-marrow-collection-of-datasets/'
ll = os.listdir(path_loc)

df_stat = pd.DataFrame()
idx = 0
for fn in ll:
    if fn[-4:] != 'h5ad': continue
    str_data_inf = fn[:-5]
    
    l = str_data_inf.split('_')
    df_stat.loc[idx,'GSE'] = l[0]
    df_stat.loc[idx,'Organism'] = l[1]
    df_stat.loc[idx,'Cells Info'] = l[2]
    df_stat.loc[idx,'Cells SubInfo'] = l[3]
    df_stat.loc[idx,'Year'] = int( l[4][1:] ); df_stat['Year'] = df_stat['Year'].astype(int)
    df_stat.loc[idx,'n_cells'] = int( l[5][:-5] ); df_stat['n_cells'] = df_stat['n_cells'].astype(int)
    df_stat.loc[idx,'Source'] = l[6]
    df_stat.loc[idx,'Counts or Preprocessed Etc'] = l[7]
    df_stat.loc[idx,'First Author'] = l[8]
    df_stat.loc[idx,'Last Author'] = l[9]
 
    t0 = time.time()
    adata = sc.read(path_loc + fn )
    df_stat.loc[idx,'Load Seconds'] = np.round(time.time()-t0,1)
    
    idx += 1
df_stat.sort_values('GSE')
    
```
