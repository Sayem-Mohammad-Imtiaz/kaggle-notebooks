```python
# RANDOM FOREST ON 'heart_disease' DATASET TO OBTAIN ACCURACY RATE AND COMPARE PREDICTED VALUE WITH THE ORIGINAL VALUE.
# ALSO CALCULATING THE ROOT MEAN SQUARED ERROR.
# HERE WE WILL TAKE COLUMNS 'age' TO 'thal'.
# ACCURACY RATE STARTS DECREASING WHEN RELATABLE COLUMNS ARE REMOVED.

import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn import model_selection
from sklearn.metrics import mean_squared_error
import numpy as np

df=pd.read_csv('../input/heart-disease-patients-details/heart_disease.csv')

Y=df['result']
X=df.loc[:,'age':'thal']

X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,random_state=7,test_size=0.33)

rf=RandomForestRegressor(n_estimators=500,oob_score=True,random_state=0) 
rf.fit(X_train,Y_train)

print(rf.score(X_train,Y_train))

prediction=rf.predict(X_test)
print(prediction)
print(Y_test)
print(np.sqrt(mean_squared_error(prediction,Y_test)))
```
