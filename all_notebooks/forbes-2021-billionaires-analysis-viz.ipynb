# Billionaires by Industry

```python
forbes_df[['Industry']].groupby(['Industry']).size()
```

```python
industry_df = forbes_df[['Industry']].groupby(['Industry']).size().reset_index(name = "counts")
industry_df.sort_values('counts', inplace = True, ascending = True)
```

finance and investments has the highest number of billionaires(371) followed by Technology with 365

```python
fig, ax = plt.subplots(figsize =(12, 7))
 
# creating the bar plot
ax.barh(industry_df['Industry'], industry_df['counts'], color ='hotPink')
# customization and cleaning up the plot
 
# Add Plot Title and axis labels

ax.set_title('Number of billionaires by Industry',fontweight = 'bold',
             loc ='left')
plt.xlabel('count of billionaires', fontweight ='bold', fontsize = 15)
plt.ylabel('Industry', fontweight ='bold', fontsize = 15)
```

# Networth by country

```python
forbes_df['NetWorth'] = forbes_df['NetWorth'].str.strip('$')
forbes_df['NetWorth'] = forbes_df['NetWorth'].str.strip('B')
```

```python
forbes_df['NetWorth'] = forbes_df['NetWorth'].astype(float)
Worth_df = forbes_df.sort_values(by = ['NetWorth'],ascending = False) 
Worth_df.head() 
```

```python
country_df=Worth_df.groupby(['Country']).size().sort_values(ascending=False).reset_index(name= 'count')
df_1=country_df[:20]
 
```

# Visualize billionaires by country

```python
fig, ax = plt.subplots(figsize =(13,6))
 
# creating the bar plot
ax.bar(df_1['Country'], df_1['count'], color ='blue')

# Add Plot Title and axis labels
ax.set_title('Number of billionaires by Country',fontweight = 'bold',
             loc ='left')
ax.set_xlabel('Country', fontweight ='bold', fontsize = 15)
ax.set_ylabel('count of Billionaires', fontweight ='bold', fontsize = 15)

#rotate xticks 45 degrees for easier readability

for tick in ax.get_xticklabels():
    tick.set_rotation(90)
```

United states has more billionaires than any country followed by china then India.

# Top 20 countries with the highest wealth 

```python
# sorting values by a multilevel index. ensure to name all columns of the index..
table1 = pd.pivot_table(Worth_df, index= 'Country', values ='NetWorth', aggfunc=['sum']).round(2)
#table1.columns
table1.sort_values(by =('sum', 'NetWorth'), ascending = False ).head(20)
```

The wealthiest individuals live in United States and have a combined networth of over 4.3 trillion dollars.
