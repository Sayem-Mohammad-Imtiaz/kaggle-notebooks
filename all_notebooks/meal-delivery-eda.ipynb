```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
train_dat = pd.read_csv("/kaggle/input/av-genpact-hack-dec2018/train.csv")
train_dat.head()
```

```python
test_dat = pd.read_csv("/kaggle/input/av-genpact-hack-dec2018/test.csv")
test_dat.head()
```

```python
meal_info_dat = pd.read_csv("/kaggle/input/av-genpact-hack-dec2018/meal_info.csv")

meal_info_dat.head()
```

```python
center_info_dat = pd.read_csv('/kaggle/input/av-genpact-hack-dec2018/fulfilment_center_info.csv')
center_info_dat.head()
```

```python
#using the information from meal_info_dat induced into train_dat

train_dat_merge = pd.merge(train_dat, meal_info_dat, how = 'inner', on = 'meal_id')
```

```python
train_dat_temp = pd.merge(train_dat_merge, center_info_dat, how = 'inner', on = 'center_id')
train_data = train_dat_temp
```

```python
train_data.head(16)
```

```python
train_data.sort_values(by=['week'])
```

```python
group_op_area_orders = train_data.groupby(['op_area'])['num_orders'].sum()
op_area_orders_gp = pd.DataFrame(group_op_area_orders)
```

```python
px.bar(op_area_orders_gp, x = op_area_orders_gp.index, y = ['num_orders'],
       title = 'OP Area vs Number of Orders')
```

```python
import numpy as np

figure_pie = px.pie(op_area_orders_gp, values = 'num_orders', names = op_area_orders_gp.index >  3.99)
figure_pie.show()

```

```python
#From above data we see week one has set of order transactions and so on it follows:

group_week_orders = train_data.groupby(['week'])['num_orders'].sum()

week_orders_gp = pd.DataFrame(group_week_orders)

figure = px.line(week_orders_gp, x = week_orders_gp.index, y = ['num_orders'], title = 'Sales Chart for all 145 Weeks')

figure.show()
```

```python
#We group on basis of region to see if one region is bring more sales over another
group_center_orders = train_data.groupby(['center_type'])['num_orders'].sum()

region_orders_gp = pd.DataFrame(group_center_orders)

px.bar(group_center_orders , x = group_center_orders.index, y = ['num_orders'], title = 'Number of Orders on Center Type')
```
