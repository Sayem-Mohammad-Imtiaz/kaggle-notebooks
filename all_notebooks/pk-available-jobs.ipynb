## it seems that all job name has [in {country_name}] at end
### lets verify

```python
last_two_words=set()
for a in df.job_name.unique().tolist():
    last_two_words.add(" ".join(a.split()[-2:]))
```

```python
last_two_words
```

#### verified

## it seems that all job start with [Full Time || Online Full time] at end
### lets verify

```python
first_few_words=set()
for a in df.job_name.unique().tolist():
    if a.startswith('Full'):
        first_few_words.add(" ".join(a.split()[:2]))
    elif a.startswith('Online'):
        first_few_words.add(" ".join(a.split()[:3]))
    else:
        first_few_words.add(" ".join(a.split()[:2]))
```

```python
first_few_words
```

#### verified

## so if we strip first few and last few words we may access more to the point data

```python
joblist = set()
for a in df.job_name.unique().tolist():
    if a.startswith('Full'):
        joblist.add(" ".join(a.split()[2:-2]))
    elif a.startswith('Online'):
        joblist.add(" ".join(a.split()[3:-2]))
```

```python
list(joblist)[0:10]
```

```python
len(joblist)
```

## it seems that all job have  [Job || Jobs] at end
### lets verify

```python
lastword = set()
for a in joblist:
    lastword.add(a.split()[-1])
```
