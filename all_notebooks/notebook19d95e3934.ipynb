# k-means clustering algo

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv("../input/pima-indians-diabetes-database/diabetes.csv")
df.head(10)
```

```python
df.drop(['Outcome'],axis='columns',inplace=True)
df.head()
```

```python
sum = []
k_rng = range(1,20)
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df)
    sum.append(km.inertia_)
    
sum
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sum)
```

```python
km = KMeans(n_clusters=2)
yp = km.fit_predict(df)
yp
```

```python
df['cluster'] = yp
df.head(10)
```

```python
df.cluster.unique()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
print(f'df1 = {df1}\n')
print(f'df2 = {df2}\n')
```

```python
km = KMeans(n_clusters = 1)
km.fit(df)
sum_k_1 = (km.inertia_)
print(sum_k_1)

km = KMeans(n_clusters = 2)
km.fit(df)
sum_k_2 = (km.inertia_)
print(sum_k_2)
```
