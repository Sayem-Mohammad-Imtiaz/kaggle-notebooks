```python
df = pd.read_excel('/kaggle/input/us-cities/uscities.xlsx')
df.head(3)
```

#New York (population)

```python
#Code by Mashrimp https://www.kaggle.com/katomash/a-car-is-moving-or-not-accuracy-94/notebook

import plotly.express as px
fig = px.scatter_mapbox(df,
                    # Here, plotly gets, (x,y) coordinates
                    lat="lat",
                    lon="lng",
                    text='county_name',

                    #Here, plotly detects color of series
                    color="population",
                    labels="state_name",

                    zoom=14.5,
                    center={"lat":40.6943, "lon":-73.9249},
                    height=600,
                    width=800)
fig.update_layout(mapbox_style='stamen-terrain')
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
fig.update_layout(title_text="New York")
fig.show()
```
