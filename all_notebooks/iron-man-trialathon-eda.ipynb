Ironman Triathlon is organized by world Triathlon Corporation. It consist of 2.4 mile swim, 112 mile bicycle ride and a marathon of 26.22 mile run, race should be in same sequence.It is considered to be one of the most toughest One Day Sports Event, this sports requires strong Mental and Fit Body so that you can surpass the toughest sport on that given date.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import warnings
import time
import scipy.stats as stats
warnings.filterwarnings('ignore')
pd.set_option('display.max_rows',2000, 'display.max_columns',100)
Iron_Man = pd.read_csv("../input/iron-man-world-champhionship-allyears/Ironman World Championship.csv",encoding= 'unicode_escape')
```

```python
Iron_Man.head()
```

```python
Iron_Man.isna().sum()
```

```python


fig = px.histogram(Iron_Man,x='Country',color="Place")
fig.update_xaxes(categoryorder="total descending",tickangle=270, tickfont=dict(size=14))
fig.update_layout(
    title='Most Winner blongs to this Countries')
fig.update_traces(marker_color='rgb(155,0,255)', marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5, opacity=0.6)
fig.show()

```

```python
Winners_Time_in_Min = []
for i in Iron_Man['Time']:
    time_obj = time.strptime(i, '%H:%M:%S')
    Winners_Time_in_Min.append(time_obj.tm_hour * 60 + time_obj.tm_min +time_obj.tm_sec/60)
Iron_Man['Winners_Time_in_Min'] = Winners_Time_in_Min
```

## Iron Man Trialathon Male Winner

```python
Iron_Man_Male = Iron_Man[Iron_Man['Gender']=='Male']
Iron_Man_Male['Year'] = Iron_Man_Male['Year'].astype(float)
fig = px.treemap(Iron_Man_Male, path=['Country','Athlete','Place','Year','Time'],hover_data=['Athlete','Place','Year','Time'],color='Country')
fig.update_layout(
    title='Iron Man Male Winiing Position-1,Position-2 and Position-3 with Detail')
fig.show()
```

## Ironman Men's Position Placewise Winner

## Place 1st for Men

```python
Iron_Man_Male_1 = Iron_Man_Male[Iron_Man_Male['Place']==1]
Iron_Man_Male_1.reset_index(inplace = True) 

Z1 = Iron_Man_Male_1['Country'].value_counts().keys()
Z2 = Iron_Man_Male_1['Country'].value_counts()

fig = go.Figure(data=[go.Scatter(
    x=Z1,
    y=Z2,
    mode='markers',
    marker=dict(
        size=[100,50,35,15,15],
        color=[100,50,35,15,15],
        showscale=True
        )
)])
fig.update_layout(
    title='Countries Winning 1st Place in Male Ironman Trialathon Race',
    xaxis_title="Countries",
    yaxis_title="Winners Count",
)
fig.show()
```

```python
fig = px.histogram(Iron_Man_Male_1,x='Athlete',color='Country')
fig.update_layout(
    title='Iron Man 1st Place Winner')
fig.update_xaxes(categoryorder="total descending",tickangle=270, tickfont=dict(size=14))
fig.show()
```

Most successful Male Ironman Trialathon winner are Dave Scott and Mark Allen winnig 6 times.
