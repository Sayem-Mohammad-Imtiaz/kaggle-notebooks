```python

import numpy as np 
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
import xgboost as xgb


```

# Loading dataset

```python
dataset = pd.read_csv("../input/mushroom-classification/mushrooms.csv")
dataset_new = pd.read_csv("../input/mushroom-classification/mushrooms.csv") #for future What if analysis
```

# Preprocessing

```python
def func (dataset):
    
    #encoding
    dataset = dataset.copy()
    dataset = dataset.apply(LabelEncoder().fit_transform)
    
    #splitting X, y
    
    y=dataset["class"].copy()
    X=dataset.drop("class", axis=1).copy()
    
    #scaling
    sc = StandardScaler()
    X = pd.DataFrame(sc.fit_transform(X)).copy()
    
    #Train, test split, i keep X for correlating and future engineering
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=123)
    
    return X, X_train, X_test, y_train, y_test
```

```python
X, X_train, X_test, y_train, y_test = func(dataset)
```

# Fitting and evaluating

```python
xg = xgb.XGBClassifier()
xg.fit(X_train, y_train)
y_pred = xg.predict(X_test)
```

```python
cf = confusion_matrix(y_test, y_pred)
```

```python
cf
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
```

```python
print('Model Accuracy : ', accuracy_score(y_test, y_pred) *  100)
print('Model Recall : ', recall_score(y_test, y_pred) *  100)
print('Model Precision : ', precision_score(y_test, y_pred) *  100)
print("F1 Score: ", f1_score(y_test, y_pred) * 100)
```

# Calculating feature importance for XGBoost

```python
importance = pd.DataFrame(xg.feature_importances_, index = name, columns = ["Score"]).sort_values("Score", ascending = False)
```

# What if section

It´s very common to get 100% recall and accuracy score on this dataset. But what if we´ll drop the most important feature, which is Gill color. Let´s find out in same fashion. 

```python
#new func without gill color
def func_new (dataset_new):
    
    #encoding
    dataset_new = dataset_new.copy()
    dataset_new = dataset_new.apply(LabelEncoder().fit_transform)
    
    #splitting X, y
    
    y2=dataset_new["class"].copy()
    X2=dataset_new.drop(["class", "gill-color"], axis=1).copy()
    
    #scaling
    sc2 = StandardScaler()
    X_new = pd.DataFrame(sc2.fit_transform(X2)).copy()
    
    #Train, test split, i keep X for correlating and future engineering
    
    X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(X_new, y2, test_size=0.20, random_state=123)
    
    return X_new, X_train_new, X_test_new, y_train_new, y_test_new
```

```python
X_new, X_train_new, X_test_new, y_train_new, y_test_new = func_new(dataset_new)
```

```python
xg2 = xgb.XGBClassifier()
xg2.fit(X_train_new, y_train_new)
y_pred_new = xg2.predict(X_test_new)
```

```python
cf_new = confusion_matrix(y_test_new, y_pred_new)
```

```python
tn_new, fp_new, fn_new, tp_new = confusion_matrix(y_test_new, y_pred_new).ravel()
```

```python
print('Model Accuracy : ', accuracy_score(y_test_new, y_pred_new) *  100)
print('Model Recall : ', recall_score(y_test_new, y_pred_new) *  100)
print('Model Precision : ', precision_score(y_test_new, y_pred_new) *  100)
print("F1 Score: ", f1_score(y_test_new, y_pred_new) * 100)
```
