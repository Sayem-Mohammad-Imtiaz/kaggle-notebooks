```python
import os
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
data0 = pd.read_csv("../input/solarpanelspower/PV_Elec_Gas3.csv")
data0[0:5].T
```

```python
fig=make_subplots(specs=[[{"secondary_y":True}]])
fig.add_trace(go.Scatter(x=data0['date'],y=data0['kWh electricity/day'].rolling(window=50).mean(),name='kWh electricity/day'),secondary_y=False,)
fig.add_trace(go.Scatter(x=data0['date'],y=data0['Gas/day'].rolling(window=50).mean(),name='Gas/day'),secondary_y=True,)
fig.update_layout(autosize=False,width=700,height=500,title_text="Solar Panel Power by kWh electricity/day and Gas/day")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="kWh electricity/day",secondary_y=False)
fig.update_yaxes(title_text='Gas/day',secondary_y=True)
fig.show()
```

```python
fig=make_subplots(specs=[[{"secondary_y":True}]])
fig.add_trace(go.Scatter(x=data0['date'],y=data0['Cumulative_solar_power'],name='Cumulative_solar_power'),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="Cumulative_solar_power")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Cumulative_solar_power",secondary_y=False)
fig.show()
```

```python
sns.jointplot(data=data0,x='kWh electricity/day',y='Gas/day',kind='hex')
plt.show()
```

```python
year=[]
month=[]
for item in data0['date']:
    items=item.split('/')
    year+=[int(items[2])]
    month+=[int(items[1])]
data0['year']=year
data0['month']=month
data0
```

```python
data_month=data0[['kWh electricity/day','Gas/day','month']].groupby('month',as_index=False).mean()
data_month
```

```python
fig=make_subplots(specs=[[{"secondary_y":True}]])
fig.add_trace(go.Scatter(x=data_month['month'],y=data_month['kWh electricity/day'],name='kWh electricity/day'),secondary_y=False,)
fig.add_trace(go.Scatter(x=data_month['month'],y=data_month['Gas/day'],name='Gas/day'),secondary_y=True,)
fig.update_layout(autosize=False,width=700,height=500,title_text="Solar Panel Power by kWh electricity/day and Gas/day by month")
fig.update_xaxes(title_text="month")
fig.update_yaxes(title_text="kWh electricity/day",secondary_y=False)
fig.update_yaxes(title_text='Gas/day',secondary_y=True)
fig.show()
```

```python
data_year=data0[['kWh electricity/day','Gas/day','year']].groupby('year',as_index=False).mean()
data_year
```

```python
fig=make_subplots(specs=[[{"secondary_y":True}]])
fig.add_trace(go.Scatter(x=data_year['year'][1:-1],y=data_year['kWh electricity/day'][1:-1],name='kWh electricity/day'),secondary_y=False,)
fig.add_trace(go.Scatter(x=data_year['year'][1:-1],y=data_year['Gas/day'][1:-1],name='Gas/day'),secondary_y=True,)
fig.update_layout(autosize=False,width=700,height=500,title_text="Solar Panel Power by kWh electricity/day and Gas/day by year(2012-2019)")
fig.update_xaxes(title_text="year")
fig.update_yaxes(title_text="kWh electricity/day",secondary_y=False)
fig.update_yaxes(title_text='Gas/day',secondary_y=True)
fig.show()
```
