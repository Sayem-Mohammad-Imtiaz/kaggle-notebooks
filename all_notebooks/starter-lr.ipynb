```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../input/car-dekho-data/car data.csv')
```

```python
df.head()
```

```python
df.Fuel_Type.replace(regex={'Petrol':'0','Diesel':'1'},inplace=True)
```

```python
df.head()
```

```python
df.drop(df.loc[df['Fuel_Type']=='CNG'].index, inplace=True)
```

```python
df.Fuel_Type.unique()
```

```python
df.Seller_Type.replace(regex={"Dealer":"0","Individual":"1"},inplace=True)
df.Transmission.replace(regex={"Manual":"0","Automatic":"1"},inplace=True)
```

```python
df[['Fuel_Type','Transmission','Seller_Type']]=df[['Fuel_Type','Transmission','Seller_Type']].astype(int)
```

```python
df.head()
```

```python
df=df[['Car_Name','Year','Present_Price','Kms_Driven','Fuel_Type','Seller_Type','Transmission','Owner','Selling_Price']]
```

```python
df
```

```python
xdat = df.drop(['Car_Name','Selling_Price'],axis=1)
```

```python
xdat
```

```python
ydat=df.Selling_Price
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test,y_train,y_test = train_test_split(xdat,ydat,test_size=0.2,random_state=4)
```

```python
lr= LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
pred=lr.predict(x_test)
```

```python
pred
```

```python
predict=pd.DataFrame(pred)
```

```python
predict
```

```python
from sklearn.metrics import r2_score
```

```python
y_test1=y_test.reset_index()['Selling_Price']
ax=y_test1.plot(label="Original Prices",figsize=(12,6),linewidth=2,color="orange")
ax=predict[0].plot(label = "Predictions",figsize=(12,6),linewidth=2,color="hotpink")
plt.legend(loc='upper right')
plt.title("Original Prices VS Predictions")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()
```

```python
r2_score(y_test,predict)
```
