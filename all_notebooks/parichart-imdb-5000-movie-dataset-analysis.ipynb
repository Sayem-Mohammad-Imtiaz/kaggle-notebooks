Pearson Correlation Interpretation:

 - What does a negative value mean? 

It means a relationship between two variables in which one variable increases as the other decreases, and vice versa.

 - What does a positive value mean? 

It means if one variable increases, the other one also increases. A positive correlation also exists in one decreases and the other also decreases.

 - What does a higher number mean versus a lower number? 

The larger the value, the stronger the relationship between the variables. On the other hand, the lower the value, the weaker the relationship between them.


```python
new_data.corr()
```

Let's compare variables to each other using pandas .corr(). It shows Pearson Correlation. The negative values indicate as one of the variables increases, the other tends to decrease while the positive values mean the relationship between variable goes together; as one variable increases or decreases, the other tends to increase or decrease with it.

```python
# Set up the matplotlib figure
plt.rcParams['figure.figsize']=[12.0, 10.0]
plt.title('Pearson Correlation')
# Draw the heatmap using seaborn
sn.heatmap(new_data.corr(),linewidths=0.5,vmax=1.0, square=True, cmap="BuGn_r", linecolor='black', annot=True)
```

For the heatmap above, we can see the correlation clearly from the color saturation. The strongest positive correlation is the relationship between actor 1 facebook likes and cast total facebook likes, which has 0.94. That means when the likes of actor 1 on facebook increase, the cast total of facebook likes increase. While the strongest negative correlation has two pairs including the relationship between title of year and duration, and IMDB score and title of year, which have -0.13.

```python
plt.rcParams['font.size']=11
plt.rcParams['figure.figsize']=[8.0, 5.5]
new_data.plot(kind='scatter', x='num_critic_for_reviews', y='movie_facebook_likes', color='Blue')
plt.title("Relationship Between Number Critic for Reviews and Movie Facebook Likes")
plt.tight_layout() 
```

Let's take a look at the scatter plots above, it shows the relationship between number of critic for reviews and movie facebook likes. The scatter plots are excellent way of looking relationship between pairs of variables In this dataset, the two variables have been created in such a way that high values on number of critic for reviews are associated with moderate value on movie facebook likes.

```python
plt.rcParams['font.size']=11
plt.rcParams['figure.figsize']=[8.0, 5.5]
new_data.plot(kind='scatter', x='num_critic_for_reviews', y='imdb_score', color='Red')
plt.title("Relationship Between Number Critic for Reviews and IMDB Score")
plt.tight_layout() 
```

For the relationship between number of critic for reviews and IMDB score, we can see from the scatter plots above. It shows that high values on number of critic for reviews are associated with high value on IMDB score.
