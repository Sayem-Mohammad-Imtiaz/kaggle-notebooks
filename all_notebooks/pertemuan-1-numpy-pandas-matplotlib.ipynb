```python
#membentuk array 2 dimensi
a = np.array([[0, 1, 2, 3],
             [10, 11, 12, 13]])
print(a)
print(a.shape) #jumlah baris, jumlah kolom
print(a.size) #baris x kolom (pada dasarnya menghitung jumlah data)
print(a.ndim)#jumlah dimensi pada array
print(a[1,3])#print array baris index ke 1, kolom index ke 3

#mengganti nilai baris index ke 1, kolom index ke 3
a[1, 3] = -1
print(a)
```

```python
#slicing
a = np.array([1, 2, 3])
#slicing dimulai dari index ke 1 dan diakhiri di index 2
print(a[1:2]) #hasil 2 karena program akan berhenti menghitung index sebelum batas index yang dituju
print(a[1:-1]) #index -1 adalah index yang dihitung dari kanan
print(a[::3]) #index awal, dan index yang tuju saja yang di print
```

```python
#PANDAS
import pandas as pd
#series
obj = pd.Series([1, 2, 3])
print(obj)#bagian kiri = index, bagian kanan = Value
print(type(obj))

#membuat series dengan index
obj2 = pd.Series([1, 2, 3], index = ["a", "b", "c"])
print(obj2)

#mendapatkan index obj2
print(obj2.index)

#print Values obj2
print(obj2.values, obj2.dtype)
```

```python
#menampilkan values pada index a
print(obj2["a"])

#assign nilai baru pada index a
obj2["a"] = 4
print(obj2)

#mengakses nilai pada index a dan c
print(obj2[["a", "c"]])
```

```python
#membuat objek series baru
obj3 = pd.Series([4, 5, 6], index = ["a", "d", "c"])
print(obj3)
print(obj2 + obj3)
```

```python
#DATAFRAME

#membuat DataFrame
data = {'kota' : ['semarang','semarang', 'semarang','bandung','bandung','bandung'],
       'tahun' : [2016, 2017, 2018, 2016, 2017, 2018],
       'populasi' : [1.5, 2.1, 3.2, 2.3, 3.2, 4.5]}
frame = pd.DataFrame(data)
print(frame)
#cek tipe
print(type(frame))
#cek shape dari frame
print(frame.shape)
```

```python
#MATPLOTLIB
#import library
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np

x = np.linspace(0,2*np.pi, 100)
cos_x = np.cos(x)

#membuat line plot
fig, ax = plt.subplots()
_ = ax.plot(x, cos_x)
```

```python
#membuat sumbu x dan y memiliki rasio yang sama
fig, ax = plt.subplots()
_ = ax.plot(x, cos_x)
_ = ax.set_aspect('equal')

#mengatur bentuk marker
fig, ax = plt.subplots()
_ = ax.plot(x, cos_x, markersize = 20, linestyle = '-.', color = 'red', label = 'cos')
_ = ax.set_aspect('equal')
_ = ax.legend()
```

```python
#membuat scatter plot
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

fig, ax =plt.subplots()
_ = ax.scatter(x, y)
_ = ax.set_xlabel('x axis')
_ = ax.set_ylabel('y axis')
```

```python
#membuat bar plot
kategori = ['Panas', 'Dingin']
jumlah = [8, 5]

fig, ax = plt.subplots()
_ = ax.bar(kategori, jumlah)
_ = ax.set_xlabel('Kategori')
_ = ax.set_ylabel('Jumlah')
_ = ax.set_title('Penikmat Kopi')
```
