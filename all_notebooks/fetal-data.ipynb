```python
y = data['fetal_health']
y = y.astype('int64')
X = data.drop(['fetal_health'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.metrics import roc_auc_score, accuracy_score, f1_score
```

```python
from sklearn.ensemble import RandomForestClassifier
trees = [140,150,160,180]
for i in trees:
    clf = RandomForestClassifier(n_estimators=i, criterion="entropy", max_depth= 5, random_state=0)
    clf.fit(X_train, y_train)
    preds = clf.predict(X_test)
    acc_sc = accuracy_score(y_test, preds, normalize=True)
    f1_sc = f1_score(y_test, preds, labels=[1,2,3], pos_label=1, average=None, zero_division='warn')
#     roc_sc = roc_auc_score(y_test, preds, average='weighted', max_fpr=None, multi_class='ovo', labels=[1,2,3])
    print("accuracy ", acc_sc)
    print("f1_score ", f1_sc)
#     print("roc_score ", roc_sc)
    print("******************************")
```

```python
X1_train = X_train.drop(['histogram_mode', 'histogram_median', 'histogram_mean'], axis=1)
X1_test = X_test.drop(['histogram_mode', 'histogram_median', 'histogram_mean'], axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
trees = [140,150,160,180]
for i in trees:
    clf = RandomForestClassifier(n_estimators=i, criterion="entropy", max_depth= 5, random_state=0)
    clf.fit(X1_train, y_train)
    preds = clf.predict(X1_test)
    acc_sc = accuracy_score(y_test, preds, normalize=True)
    f1_sc = f1_score(y_test, preds, labels=[1,2,3], pos_label=1, average=None, zero_division='warn')
#     roc_sc = roc_auc_score(y_test, preds, average='weighted', max_fpr=None, multi_class='ovo', labels=[1,2,3])
    print("accuracy ", acc_sc)
    print("f1_score ", f1_sc)
#     print("roc_score ", roc_sc)
    print("******************************")
```

```python
from sklearn.ensemble import RandomForestClassifier
trees = [140,150,160,180]
for i in trees:
    clf = RandomForestClassifier(n_estimators=i, criterion="gini", max_depth= 5, random_state=0)
    clf.fit(X1_train, y_train)
    preds = clf.predict(X1_test)
    acc_sc = accuracy_score(y_test, preds, normalize=True)
    f1_sc = f1_score(y_test, preds, labels=[1,2,3], pos_label=1, average=None, zero_division='warn')
#     roc_sc = roc_auc_score(y_test, preds, average='weighted', max_fpr=None, multi_class='ovo', labels=[1,2,3])
    print("accuracy ", acc_sc)
    print("f1_score ", f1_sc)
#     print("roc_score ", roc_sc)
    print("******************************")
```

```python
from sklearn.neighbors import KNeighborsClassifier

neighbors = [1,3,5,7,9]

for i in neighbors:
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    
    predsk = knn.predict(X_test)
    
#     predsk = (np.argmax(knn.predict(X_test)))+1
    acc_sc = accuracy_score(y_test, predsk, normalize=True)
    f1_sc = f1_score(y_test, predsk, labels=[1,2,3], pos_label=1, average=None, zero_division='warn')
# #     roc_sc = roc_auc_score(y_test, preds, average='weighted', max_fpr=None, multi_class='ovo', labels=[1,2,3])
    print("accuracy ", acc_sc)
    print("f1_score ", f1_sc)
#     print("roc_score ", roc_sc)
#     print(predsk)
    print("******************************")
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components=2)

lda_data_mod = lda.fit_transform(X,y)
lda_data = pd.DataFrame(data=lda_data_mod, columns=['C1', 'C2'])

lda_data = pd.concat([lda_data, y], axis=1)
sns.lmplot(x='C1', y='C2', data=lda_data, hue='fetal_health', fit_reg=False)
```

```python
lda.explained_variance_ratio_
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(X,y)
pca_data = pd.DataFrame(data=pca_data, columns=['P1', 'P2'])

pca_data = pd.concat([pca_data, y], axis=1)
sns.lmplot(x='P1', y='P2', data=pca_data, hue='fetal_health', fit_reg=False)
```

```python
pca_lda = pd.concat([pca_data.drop(['fetal_health'], axis=1), lda_data], axis=1)
pca_lda.head()
```

```python
from sklearn.metrics import precision_score, recall_score, confusion_matrix, classification_report, accuracy_score, f1_score
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier(subsample=0.9, max_features=0.7, n_estimators=150, random_state=42)
gbc.fit(X_train, y_train)
stochastic_boost_score = gbc.score(X_test, y_test)
print('accuracy is: ', stochastic_boost_score*100)
y_pred_gbc = gbc.predict(X_test)
f1_sc = f1_score(y_test, y_pred_gbc, average='micro')
f1_sc
```

```python
from sklearn.metrics import accuracy_score,plot_confusion_matrix,classification_report, matthews_corrcoef

print("Accuracy score: {}".format(accuracy_score(y_test,y_pred_gbc)))
print("Classification report:\n {}".format(classification_report(y_test,y_pred_gbc)))
print("Matthew Correlation Coefficient Score: {}".format(matthews_corrcoef(y_test,y_pred_gbc)))
plot_confusion_matrix(gbc, X_test, y_test, cmap='Blues')
```
