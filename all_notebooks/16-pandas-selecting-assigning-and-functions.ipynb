```python
square_a = lambda x: x*x
print(square_a(3))
```

```python
# 보험증서 개수별로 LTV 최대값을 찾음
df.groupby('Number of Policies')['Customer Lifetime Value'].max()
```

```python
# apply 사용법 (1) - 데이터프레임을 받아 LTV 값이 최대인 행을 출력하는 람다합수를 적용

df.groupby('Number of Policies').apply(lambda df: df.loc[df['Customer Lifetime Value'].idxmax()])  # idxmax()는 최대값을 가진 행 인덱스를 출력하는 함수
```

```python
# apply 사용법 (2) - Months Since Last Claim 컬럼에 해당 람다함수를 적용해 새로운 열 생성
df['Years Since Last Claim'] = df['Months Since Last Claim'].apply(lambda x: '1 Year' if 12<=x<24 else ('2 Years' if 24<=x<36 else 'Below 1 Year'))
print(df['Years Since Last Claim'])
```

```python
train = pd.read_csv('../input/data-science-day1-titanic/DSB_Day1_Titanic_train.csv')
```

```python
train
```

```python
#나이값의 빈 값의 개수
train.Age.isnull().sum()
```

```python
# Name 컬럼에서 .앞에 이니셜 문자 추출(Mr. Mrs. Miss. 등)해 새로운 Initial 컬럼 생성
train['Initial']=train.Name.str.extract('([A-Za-z]+)\.')
```

```python
train
```

```python
# apply 사용법 (3) - 나이 컬럼의 빈값을 이니셜별 나이 평균 값으로 채우기
train.Age = train.groupby('Initial').Age.apply(lambda x: x.fillna(x.mean()))
```
