```python
import pandas as pd
tweet=pd.read_csv('../input/farmers-protest-tweets-dataset-csv/tweets.csv',
                  usecols=["date", "renderedContent"])
tweet.head()
```

```python
stock=pd.read_csv('../input/sp-bse-sensex/CSVForDate (3).csv')
stock.head()
```

```python
stock=stock[['Date','Close']]
stock.head()
```

```python
stock['Price Change']=stock['Close'].diff(periods=1)
stock.head()
```

```python
stock = stock[stock['Price Change'].notna()]
stock.head()
```

```python
def getAnalysis(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'
stock['Sentiment'] = stock['Price Change'].apply(getAnalysis )
stock.head()
```

```python
stock=stock[['Date','Sentiment']]
stock.head()
```

```python
stock.Sentiment.value_counts()
```

```python
import datetime as dt
stock.Date=pd.to_datetime(stock.Date)
stock.head()
```

```python
tweet[['Date','Time']] = tweet.date.str.split(expand=True)
tweet.head()
```

```python
tweet=tweet[['Date','renderedContent']]
tweet.head()
```

```python
tweet.Date=pd.to_datetime(tweet.Date)
tweet.head()
```

```python
data=tweet.merge(stock, on='Date', how='inner')
data.isnull().sum()
```

```python
data=data.dropna()
data.isnull().sum()
```

```python
data.head()
```

```python
round(data.Sentiment.value_counts()*100/len(data),2)
```

```python
data.to_csv(r'./mbsa.csv', index = False)
```
