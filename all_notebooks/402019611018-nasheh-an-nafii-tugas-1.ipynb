# A. Load Data

```python
data = pd.read_csv("/kaggle/input/harga-real-estate/harga_real_estate.csv")
data.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

#menggunakan metode SkalaMinMax
scaler = MinMaxScaler()
df = pd.DataFrame(data) #convert dengan dataframe
df = scaler.fit_transform (df)

#membuat data menjadi tabel
df = pd.DataFrame(data=df,columns=['Y_harga_unit','X2_jarak_MRT','X3_jumlah_toko','X1_usia_Rumah','No'])

a = df.drop(['No'],axis=1) #mengambil data input
X = a.drop(['X1_usia_Rumah'],axis=1)
y = df.iloc[:,-2] #mengambil data output
print(X,y)
```

```python
from sklearn.model_selection import train_test_split

#membagi data train dan data test
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
print(lr.coef_)
```

```python
pred_y = lr.predict(X_test)

print("The first five prediction {}".format(pred_y[:5]))
print("The real first five labels {}".format(y_test[:5]))

mse = lr.score(X_test, y_test)
print("R2 {}".format(mse))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(lr, X_train, y_train, scoring='r2', cv=10)
print(scores)
```
