```python
data=pd.read_csv('/kaggle/input/diamonds/diamonds.csv',index_col=0)
```

```python
data
```

```python
data.isna().any()
```

```python
X=pd.get_dummies(data,drop_first=True)
```

```python
y=X.pop('price')
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.svm import SVR
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values, test_size=0.33, random_state=42)
```

```python
reg1=SVR(kernel='poly')
reg1.fit(X_train,y_train)
```

```python
r2_score(y_test,reg1.predict(X_test))
```

```python
reg1.predict(X_test)
```

```python
y_test
```

```python
from sklearn.ensemble import GradientBoostingRegressor,AdaBoostRegressor, RandomForestRegressor
```

```python
reg2 = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1,max_depth=1, random_state=0, loss='ls').fit(X_train, y_train)
```

```python
reg2.predict(X_test)
```

```python
y_test
```

```python
r2_score(y_test,reg2.predict(X_test))
```

```python
reg3=AdaBoostRegressor().fit(X_train,y_train)
```

```python
r2_score(y_test,reg3.predict(X_test))
```

```python
reg4=RandomForestRegressor().fit(X_train,y_train)
```

```python
y_pred=reg4.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```
