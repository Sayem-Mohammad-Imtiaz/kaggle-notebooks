```python
df_wgt_user = pd.read_csv('../input/lowcol-sample007-fet2-randinit01/out_user.csv', index_col=0)
print(df_wgt_user.shape)
df_wgt_user.head()
```

```python
df_wgt_col = pd.read_csv('../input/lowcol-sample007-fet2-randinit01/out_col.csv', index_col=0)
print(df_wgt_col.shape)
df_wgt_col.head()
```

```python
loadings = pd.read_csv('../input/apply-factanal-to-sample007-cor/loadings.csv', index_col=0)
print(loadings.shape)
loadings.head()
```

```python
scores = df.values.dot(loadings.values)
print(scores.shape)
```

```python
'''
排他でクラスづけする
値の一番高いところを該当クラスとする
'''
scores.argmax(axis=1)
```

```python
'''うまくクラスタリングできている'''
df1 = df_wgt_user.copy()
df1['cls'] = ['cls' + str(ee) for ee in scores.argmax(axis=1)]
df1
plt.figure(figsize=(10,10))
g = sns.lmplot(data=df1, x='0', y='1', fit_reg=False, hue='cls')
save_img()
```
