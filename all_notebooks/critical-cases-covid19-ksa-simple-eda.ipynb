```python
# import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import datetime
import time
```

```python
# load data.
data = pd.read_csv('../input/api-data-critical-cases-ksa-covid19/api_critical_cases_ksa_covid19.csv')
data.head()
```

```python
# Each row is a critical case 
data= data.set_index('ObjectId')
```

```python
# visualize null values.
fig, ax = plt.subplots(figsize = (10, 5))
sns.heatmap(data.isnull(), yticklabels=False, ax = ax, cbar=False, cmap='viridis')
ax.set_title('Data')
```

```python
data.info() # check columns' name and type.
```

```python
data.OBJECTID_1.isna().sum() # OBJECTID_1 has 28481 null values, drop that column.
```

```python
data= data.drop('OBJECTID_1', axis=1) # drop OBJECTID_1 column.
```

```python
# statistics.  
data.describe() # outlier at least in max level.
# Tested has outlier at least in max and min level.
```

```python
# correlation between features.
fig, ax = plt.subplots(figsize = (16,16))
sns.heatmap(data.corr(),annot= True, cmap = 'coolwarm')
ax.set_title('Correlation Between Features')
```

```python
data_d= data.drop(columns=['PlaceName_AR','GovernorateName_AR','RegionName_AR','Name_Ar'], axis=1)  # drop arabic columns.
# convert epoch to human-readable date
data_d['Date'] = data_d['Reportdt']/1000 
data_d['Date']= data_d['Date'].apply(lambda x: time.strftime("%Y-%m-%d %H:%M:%S", time.gmtime(x)))

data_d.head()
```
