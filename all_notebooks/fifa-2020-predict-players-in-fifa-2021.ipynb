```python
data15 = pd.read_csv("/kaggle/input/fifa-20-complete-player-dataset/players_15.csv")
data16 = pd.read_csv("/kaggle/input/fifa-20-complete-player-dataset/players_16.csv")
data17 = pd.read_csv("/kaggle/input/fifa-20-complete-player-dataset/players_17.csv")
data18 = pd.read_csv("/kaggle/input/fifa-20-complete-player-dataset/players_18.csv")
data19 = pd.read_csv("/kaggle/input/fifa-20-complete-player-dataset/players_19.csv")
data20 = pd.read_csv("/kaggle/input/fifa-20-complete-player-dataset/players_20.csv")
data = [data15,data16,data17,data18,data19,data20]
```

## Number of players during (2015-2020) in FIFA

```python
import plotly.express as px

years = [i for i in range(2015,2021)]
players = [data15.shape[0], data16.shape[0], data17.shape[0], data18.shape[0], data19.shape[0], data20.shape[0]]
df = pd.DataFrame()
df['year'] = years
df['players'] = players

fig = px.line(df, x='year', y='players')
fig.show()
```

## Info about players since 2015

```python
def common_players(data=[data15,data16,data17,data18, data19, data20], by="short_name"):
    a = set(data[0][by])
    for i in range(1,len(data)):
        a = a.intersection( set(data[i][by]) )    
    return a

print("Number of players since 2015:",len(common_players()))
```

## Top nationatilies (2015-2020)

```python
import plotly.graph_objects as go

top2015 = list(data15['nationality'].value_counts()[0:10].index)

fig = go.Figure()
for i in top2015:
    a = []
    for j in data:
        a.append(j['nationality'].value_counts()[i])
    fig.add_trace(go.Scatter(x=years, y=a,
                    mode='lines+markers',
                    name=i))
fig.show()
```

```python
for i in range(0,6):
    data[i]['age_bin'] = data[i]['age'].apply(lambda x: "[0-21]" if x <=21 else ("[22-25]" if x>21 and x <=25 else 
                                                                                 ("[26-30]" if x>26 and x <=30 else "[31-45]")))

fig = go.Figure(data=[
    go.Bar(name='[0-21]', x=years ,y=[i['age_bin'].value_counts()["[0-21]"] for i in data]),
    go.Bar(name="[22-25]", x=years ,y=[i['age_bin'].value_counts()["[22-25]"] for i in data]),
    go.Bar(name="[26-30]", x=years ,y=[i['age_bin'].value_counts()["[26-30]"] for i in data]),
    go.Bar(name='[31-45]', x=years ,y=[i['age_bin'].value_counts()["[31-45]"] for i in data]),
])

fig.update_layout(barmode='stack')
fig.show()
```
