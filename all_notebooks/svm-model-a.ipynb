```python
# all the packages needed
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.pipeline import Pipeline
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
import seaborn as sns
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        locat=os.path.join(dirname, filename)
```

```python
# importing the dataset
dataset = pd.read_csv(locat)
# taking the values from the dataframe
data1 = dataset.values
# Check if there is an empty cell
w = np.where(np.isnan(data1))
print(w)
# it looks like we have empty cells in 16 locations
# so one easy way to deal with this is to drop them
r = w[0] # assign the locations of the rows 
data1 = np.delete(data1,r,0)
print(np.where(np.isnan(data1))) 
# it looks we solved it 
# Seperat the inputs and the output
# notice that the ID column doesn't affect the output, so it will be droped 
x = data1[:,1:-1]
y = data1[:,-1]
```

```python
# split the dataset into train and test sets
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.3)
# To see if the first two components of PCA are big enough
pca = PCA(n_components=2)
pca.fit(x_train)
# Standardize the inputs
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_train)
x_p = pca.transform(x_scaled)
colormap = np.array(['r','b'])
y_train = y_train.astype(int) # change array from float to integer, to use it as indcies 
sns.set()
plt.scatter(x_p[:,0],x_p[:,1], c=colormap[y_train]) # plot the pca 
```

```python
pipe = [('Standardize',StandardScaler()),('PCA',PCA(n_components=2)),('SVM',SVC(kernel='rbf'))]
classify = Pipeline(pipe)
classify.fit(x_train,y_train)
yhat = classify.predict(x_test)
accuracy = accuracy_score(y_test,yhat)
print('Accuracy: %.3f' % accuracy)
```
