```python
df.drop(columns=['SpType'], axis=1, inplace=True)
```

**Seperating data and target**

```python
X = df.drop(columns=['TargetClass'])
y = df['TargetClass']
```

```python
print(X, X.shape)
```

```python
print(y, y.shape)
```

**Split into train and test**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
print(X_train.shape, X_test.shape)
```

**Standardizing data**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
print(X_train)
```

**Logistic Regression model training**

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
```

```python
classifier.fit(X_train, y_train)
```

**Evaluating model**

```python
from sklearn.metrics import accuracy_score
```

```python
train_class = classifier.predict(X_train)
print(train_class)
```

```python
accuracy = accuracy_score(y_train, train_class)
print('Accuracy score: ', accuracy)
```

```python
## Test data
test_class = classifier.predict(X_test)
print(test_class)
```

```python
accuracy_test = accuracy_score(y_test, test_class)
print('Accuracy score: ', accuracy_test)
```
