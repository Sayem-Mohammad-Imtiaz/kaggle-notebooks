```python
df2= df1[df1.columns[df1.isna().sum()/df1.shape[0] < 0.7] ]
    
    # 1/ df.isna().sum()/df.shape[0] >0.7     2/df.columns[df.isna().sum()/df.shape[0] > 0.7]

    # What we do here is boolean indexing, i.e. I take the table of columns df.columns[df.isna().sum()/df.shape[0] < 0.7] 
    # and I inject it in our new dataframe df and we save 
```

## The number of people having the word captain in their job title 

```python
def captain(string):
    if 'captain' in (string.lower()):
        return True
    else:
        return False
```

```python
 captain('CAPTAIN SERVICE') # test of function 
```
