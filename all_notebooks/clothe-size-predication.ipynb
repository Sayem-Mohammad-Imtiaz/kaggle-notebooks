****Age and height have few blank values, let clean the data

```python
data['age'].replace('', np.nan, inplace=True)
data['height'].replace('', np.nan, inplace=True)
data.dropna(subset=['height'], inplace=True)
data.dropna(subset=['age'], inplace=True)
print(data)
```

```python
data.head()
v=data.replace(to_replace=["XXS","S","M","L","XL","XXL","XXXL"],value=["1","2","3","4","5","6","7"])
```

```python
v.head()
```

```python
X = v[['weight','age','height']]
y = v['size']
regr = linear_model.LinearRegression()
regr.fit(X, y)
```

```python
predict_size = regr.predict([[68,23,156]])
print(predict_size)
```
