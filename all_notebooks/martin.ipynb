```python
#导入模块
import pandas as pd 
import numpy as np
import  matplotlib.pyplot as plt
import seaborn as sns
import pymc3 as pm
```

```python
#导入数据并初步检查
Card = pd.read_csv('../input/credit-card-customers/BankChurners.csv')
Card.head(10)
Card.columns
Card.index
```

```python
#根据字段名称判断,去掉最后两列
Card = Card.drop(['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1','Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'],axis=1)
pd.set_option('display.max_columns', 30)
pd.set_option('display.max_rows', 10)
pd.set_option('display.width', 200)
Card.head(10)
Card.columns
```

```python
#进一步查看数据的基本情况
Card.describe()
len(Card)
Card.isnull().sum()
```

```python
#检查是否有重复值
Card[Card.duplicated()]
```

```python
#数据特征
numeric_features = Card.select_dtypes(include=[np.number])
print(numeric_features.columns)
```

```python
#类别特征
categorical_features = Card.select_dtypes(include=[np.object])
print(categorical_features.columns)
```

```python
#查看类别型变量的数值内容
for i in list(categorical_features.columns):
    print('--------------------------------')
    print('{0}+的唯一值如下：{1}'.format(i,Card[i].nunique()))
    print(Card[i].unique())
```

```python
#查看数值变量类型分布特征
plt.figure(figsize=(30,120))
col = list(numeric_features.columns)
for i in range(1,len(numeric_features.columns)+1):
    ax=plt.subplot(9,2,i) 
    sns.kdeplot(Card[col[i-1]])
    plt.xlabel(col[i-1])

plt.show()  
```

```python
#看下数值型变量的唯一值  
for i in list(numeric_features.columns):
    print('{0}+的唯一值如下：{1}'.format(i,Card[i].nunique()))
    
#根据唯一值数量多寡,将数值变量类型进行分类  
num_cate_col = [  'Dependent_count', 'Total_Relationship_Count', 'Months_Inactive_12_mon','Contacts_Count_12_mon' ]
numeric_col = [ x for x in list(numeric_features.columns) if x not in num_cate_col]
numeric_col.remove('CLIENTNUM')
numeric_col
```

```python
#对num_cate_col进行描述性分析
for i in num_cate_col:
    print('-------------------------------------------')
    print('{0}+的唯一值如下：{1}'.format(i,Card[i].nunique()))
    print(Card[i].unique())

Card['Attrition_Flag'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%')

cate_cold = list(categorical_features.columns)+num_cate_col
for i in cate_cold:
    tmp = pd.crosstab(Card[i], Card['Attrition_Flag'])
    tmp['总人数'] = tmp.sum(axis=1)
    tmp['流失率'] = tmp['Attrited Customer']/tmp['总人数']
    fig, ax1 = plt.subplots()
    ax1.bar(tmp.index,tmp['总人数'],color='blue')
    ax2 = ax1.twinx() 
    ax2.plot(tmp.index,tmp['流失率'],color='red')
    plt.show()
```
