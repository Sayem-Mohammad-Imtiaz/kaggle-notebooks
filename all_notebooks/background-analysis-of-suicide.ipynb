Here we will try to understand differnt attributes of the dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1= pd.read_csv('../input/suicide-rates-overview-1985-to-2016/master.csv')
```

1. Countries with highest population

```python
fig, axes = plt.subplots(figsize=(14, 10))
df2 = df1.sort_values(by='population', ascending=False)
s=sns.barplot(data=df2, x='population', y='country', palette='CMRmap', ci=None)
s.set_xlabel('Country')
s.set_ylabel('Population')
s.set_title('Top 10 Countries with Highest Population', fontsize=15)
s.set_ylim([9, 0])
plt.show()
```

2. Country with highest GDP per capita:

```python
plt.figure(figsize=(12,6))
df_sorted = df1.sort_values('gdp_per_capita ($)', ascending=False)
g = plt.bar('country', 'gdp_per_capita ($)', data=df_sorted)
plt.xticks(rotation=45)
plt.xlim((0, 9))
plt.show()
```

```python
fig, axes = plt.subplots(figsize=(14, 10))
order = df1.sort_values(by='gdp_per_capita ($)', ascending=False)
s = sns.barplot(data=order, x='country', y='gdp_per_capita ($)', ci= None)
for p in s.patches:
    s.annotate(format(p.get_height(), '.1f'),
               (p.get_x() + p.get_width() / 2., p.get_height()),
               ha='center', va='top',
               xytext=(0, 10), size=9,
               textcoords='offset points')
axes.set_xlim([9, 0])
plt.show()
```
