```python
def getData(path):
    anger = 0
    fear = 0
    sad = 0
    happy = 0
    surprise = 0
    neutral = 0
    df = pd.read_csv(path)
    
    X=[]
    y=[]
    
    for i in range(len(df)):
        if df.iloc[i]['emotion'] !=1:
            if df.iloc[i]['emotion'] == 0:
                if anger <= 4000:
                    y.append(df.iloc[i]['emotion'])
                    im = df.iloc[i]['pixels']
                    im = [int(X) for X in im.split()]
                    X.append(im)
                    anger = anger+1
                
            else:
                pass
            
           
            
            if df.iloc[i]['emotion'] == 2:
                if fear <= 4000:
                    y.append(df.iloc[i]['emotion'])
                    im = df.iloc[i]['pixels']
                    im = [int(X) for X in im.split()]
                    X.append(im)
                    fear = fear+1
            else:
                pass
            
            if df.iloc[i]['emotion'] == 3:
                if happy <= 4000:
                    y.append(df.iloc[i]['emotion'])
                    im = df.iloc[i]['pixels']
                    im = [int(X) for X in im.split()]
                    X.append(im)
                    happy = happy+1
            else:
                pass
            
            if df.iloc[i]['emotion'] == 4:
                if sad <= 4000:
                    y.append(df.iloc[i]['emotion'])
                    im = df.iloc[i]['pixels']
                    im = [int(X) for X in im.split()]
                    X.append(im)
                    surprise = surprise+1
            else:
                pass
            
            if df.iloc[i]['emotion'] == 5:
                if sad <= 4000:
                    y.append(df.iloc[i]['emotion'])
                    im = df.iloc[i]['pixels']
                    im = [int(X) for X in im.split()]
                    X.append(im)
                    sad = sad+1
            else:
                pass
            
            if df.iloc[i]['emotion'] == 6:
                if neutral <= 4000:
                    y.append(df.iloc[i]['emotion'])
                    im = df.iloc[i]['pixels']
                    im = [int(X) for X in im.split()]
                    X.append(im)
                    neutral = neutral+1
            else:
                pass
            
    return X, y
            
             
```

```python
X, y = getData('../input/facial-expression/fer2013.csv')
```

```python
np.unique(y, return_counts =True)
```

```python
X = np.array(X)/255.0
y = np.array(y)
```

```python
X.shape, y.shape
```

```python
y_o = []
for i in y:
    if i!=6:
        y_o.append(i)
    
    else:
        y_o.append(1)
```

```python
np.unique(y_o, return_counts=True)
```

```python
X = X.reshape(len(X), 48, 48, 1)
```

```python
# no. of images, height, width, color map
```

```python
X.shape
```

```python
from keras.utils import to_categorical
y_new = to_categorical(y_o, num_classes=6) #converting to 2d vector
```

```python
len(y_o), y_new.shape
```

```python
y_o[20], y_new[20]
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout, Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.metrics import categorical_accuracy
from keras.models import model_from_json
from keras.callbacks import ModelCheckpoint
from keras.optimizers import *
from keras.layers.normalization import BatchNormalization
```

```python
model = Sequential()
input_shape = (48,48,1)

model.add(Conv2D(64, (5,5), input_shape = input_shape, activation = 'relu', padding = 'same'))
model.add(Conv2D(64, (5,5), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(128, (5,5), activation = 'relu', padding = 'same'))
model.add(Conv2D(128, (5,5), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Conv2D(256, (3,3), activation = 'relu', padding = 'same'))
model.add(Conv2D(256, (3,3), padding='same'))
model.add(BatchNormalization())
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2,2)))

model.add(Flatten())
model.add(Dense(6, activation='softmax'))

model.compile(loss = 'categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
model.fit(X, y_new, epochs=22, batch_size=64, shuffle=True, validation_split=0.2)
```

```python
model.save('model.h5')
```
