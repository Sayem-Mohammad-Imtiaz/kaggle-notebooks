```python
dt = pd.read_csv('/kaggle/input/solar-flare-list-over-12-years/SolarFlare_list_GOES.csv', index_col=0)
dt.head()
```

```python
dt = dt[dt['AR number']> 0]
dt.info()
```

```python
plt.rcParams['figure.figsize'] = (20,10)
# sns.distplot(dt['GOES Duration, s'], color='red',bins=13)
# sns.histplot(dt['GOES Duration, s'], color='red',bins=13)
# time duration of 5 minutes 300 seconds
sns.histplot(dt['GOES Duration, s'], color='red',binwidth=300)
plt.xlabel('Duration time in seconds', size=15)
```

```python
plt.figure(figsize = (12, 8))
corr_time = dt.corr()
sns.heatmap(corr_time, annot = True);
```

```python
ndt = dt.loc[(dt['GOES Class'] > "C1.0") & (dt['GOES Class'] < "M1.0"),:]
sns.histplot(ndt['GOES Duration, s'], color='red',binwidth=300)
```
