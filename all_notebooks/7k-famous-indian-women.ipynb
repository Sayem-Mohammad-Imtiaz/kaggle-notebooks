```python
# Read CSV file

ind_womens = pd.read_csv('/kaggle/input/7k-indian-famous-women/data.csv')

ind_womens.head()
```

Data Cleaning

```python
#Find duplicate Values

ind_womens.duplicated().sum()
```

```python
#Drop Dupliate Value

ind_womens.drop_duplicates()
```

```python
#check null values in data set

ind_womens.isna().sum()
```

```python
#Drop Coloums

ind_womens.drop(['Father','Mother','Spouse'],inplace=True, axis=1)
```

```python
#Replace NAN Value with 0

ind_womens.fillna(0)
```

**Women in differnt Job Roles**

```python
plt.figure(figsize=(14,8))

sns.countplot(y= ind_womens['Job'], order=ind_womens['Job'].value_counts().index[:20])

plt.xlabel('Number of womens')
```

**Womens belong country**

```python
#Find higest country value
plt.figure(figsize=(14,8))

sns.countplot(y= ind_womens['Country'], order=ind_womens['Country'].value_counts().index[:20])

plt.xlabel('Country Name')
```
