Uma métrica interessante para esse dataset é a acurácia.
A acurácia mede com qual frequência o classificador faz a predição correta.

Ela é dada pela divisão entre as predições corretas do modelo (Verdadeiro positivo e Falso verdadeiro) por todas as predições do modelo.

* Verdadeiro positivo - TP
* Falso positivo - FP
* Verdadeiro negativo - TN
* Falso negativo - FN

Assim a acurácia é:
$acc = \frac{TP + TN}{TP + TN + FP+ FN }$

Com a acurácia podemos saber o quanto o modelo acertou e assim melhorá-lo.

Seria interessante aplicar um modelo de clusterização para o conjunto de variáveis:
* 'Age' e 'Spending Score (1-100)'
* 'Annual Income' e 'Spending Score (1-100)'
* 'Age', 'Annual Income' e 'Spending Score (1-100)'

Podemsos usar as métricas de Homogeneidade e Completude para avaliar os clusters.

Na primeira, um cluster perfeitamente homogêneo é aquele em que cada cluster possui pontos de dados pertencentes ao mesmo rótulo de classe. 
A homogeneidade descreve a proximidade do algoritmo de agrupamento a essa perfeição.

Na segunda, um cluster perfeitamente completo é aquele em que todos os pontos de dados pertencentes à mesma classe estão agrupados no mesmo cluster. A integridade descreve a proximidade do algoritmo de agrupamento a essa perfeição.
