```python
##loading the data frames into variables for usage

#2020
april_20_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2020/April2020.csv", parse_dates=True)
august_20_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2020/August2020.csv", parse_dates=True)
december_20_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2020/December2020.csv", parse_dates=True)
february_20_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2020/February2020.csv", parse_dates=True)
january_20_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2020/January2020.csv", parse_dates=True)
july_20_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2020/July2020.csv", parse_dates=True)
june_20_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2020/June2020.csv", parse_dates=True)
march_20_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2020/March2020.csv", parse_dates=True)
may_20_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2020/May2020.csv", parse_dates=True)
november_20_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2020/November2020.csv", parse_dates=True)
october_20_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2020/October2020.csv", parse_dates=True)
september_20_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2020/September2020.csv", parse_dates=True)

#2021
april_21_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2021/April2021.csv", parse_dates=True)
february_21_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2021/February2021.csv", parse_dates=True)
january_21_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2021/January2021.csv", parse_dates=True)
march_21_df = pd.read_csv("../input/citibike-bike-sharingnewyork-cityjan-to-apr-2021/CitiBike_Trip_Data/2021/March2021.csv", parse_dates=True)

#merging into a single dataframe
stacked_df =  pd.concat([april_20_df,august_20_df,december_20_df,february_20_df,january_20_df,july_20_df,june_20_df,
                         march_20_df,may_20_df,november_20_df,november_20_df,october_20_df,september_20_df,april_21_df,
                         february_21_df,january_21_df,march_21_df], axis = 0)
citibike_df = stacked_df.reset_index(drop = True)
```

# Since the data looks clean and usable, lets do some data analysis and visualizations

the description of the columns are as follows:
- tripduration - Duration in Seconds
- starttime - Start Time and Date
- stoptime - Stop Time and Date
- start station id - ID of Start Station
- start station name - Name of Start Station
- start station latitude - Latitude of start station
- start station longitude - Longitude of start station
- end station id - ID of End Station
- end station name - Name of End Station
- end station latitude - Latitude of End station
- end station longitude - Longitude of End station
- Bike ID - Bike ID
- usertype - Customer = 24-hour pass or 3-day pass user; Subscriber = -Annual Member
- gender - Zero=unknown; 1=male; 2=female
- birthyear - Year of Birth

# subscribers  vrs non-subscribers

```python
plt.figure(figsize=(20,15))
sns.catplot(x = citibike_df['usertype'], y = citibike_df['tripduration'],
            jitter = True, data = citibike_df);
```

Based off of the readme discription of usertype, we can see that the vast majority of time spent on trips are from people that purchase a 24 hour or 3 day pass as opposed to those with a yearly membership. We can also see that the majority of trips taken across both types are commonly less than 30 minutes.

# plot time series

```python
# changing object to datetime and grouping into a dataframe
#date_df = pd.to_datetime(citibike_df['starttime'], infer_datetime_format=True)
citibike_df = citibike_df.rename(columns = {'starttime' : 'datetime'})
citibike_df['datetime'] = pd.to_datetime(citibike_df['datetime'])
citibike_df['month'] = citibike_df['datetime'].dt.month
citibike_df['day'] = citibike_df['datetime'].dt.day
citibike_df['year'] = citibike_df['datetime'].dt.year
citibike_df['hour'] = citibike_df['datetime'].dt.hour
citibike_df
```

# busiest time of day

```python
rentals_per_hour = pd.value_counts(citibike_df['hour'])
rph_df = rentals_per_hour.to_frame()
rph_df.reset_index(inplace=True)
rph_df.rename(columns = {'index' : 'time', 'hour' : 'rides'},inplace=True)
```

```python
plt.figure(figsize=(20,5))
sns.lineplot(x = "time", y = "rides",data=rph_df);
```

we can see the busiest time of day peaks around 5-6pm with a steap drop off by 8pm
