```python
def airbag(x):
    if x=='airbag':
        return 1
    else:
        return 0
```

```python
data_train['airbag']=data_train['airbag'].apply(airbag)
```

```python
def seatbelt(x):
    if x=='belted':
        return 1
    else:
        return 0
```

```python
data_train['seatbelt']=data_train['seatbelt'].apply(seatbelt)
```

```python
def dead(x):
    if x=='dead':
        return 1
    else:
        return 0
```

```python
data_train['dead']=data_train['dead'].apply(dead)
```

```python
def fun(x):
    if x=='m':
        return 1
    else:
        return 0
```

```python
data_train['sex']=data_train['sex'].apply(fun)
```

```python
def occRole(x):
    if x=='driver':
        return 1
    else:
        return 0
```

```python
data_train['occRole']=data_train['occRole'].apply(occRole)
```
