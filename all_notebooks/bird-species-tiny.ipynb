creating the required directories

```python
def mkdirs(path, dirs):
    path.mkdir(parents=False, exist_ok=True)
    for dir in dirs:
        dir = path/dir
        dir.mkdir(parents=True, exist_ok=True)
```

```python
def replace(path):
    return path.replace('\\', '/')
```

# Correcting the file path separator in the final csv and saving it.

```python
df_filtered['filepaths'] = df_filtered.filepaths.map(replace)
df_filtered.head(4)
```

```python
df_filtered.to_csv(tiny/'bird-species-tiny.csv')
```

```python
def copy_files(src_root: Path, srcs: List, dest_root: Path):
    for src in srcs:
        shutil.copy(src_root/src, dest_root/src)
```

```python
Path(df.filepaths[0]).parent.name
```

```python
'train'/Path(bird_a)
```

```python
fns = df_filtered['filepaths'].to_list()
birds = [bird_a, bird_b]
selected_birds = [f for f in fns if Path(f).parent.name in birds]
selected_birds[:10]
```

# Copying the selected birds from the original dataset into the tiny dataset

```python
copy_files(orig, selected_birds, tiny)
```

# Creating the final zip file

```python
shutil.make_archive('bird-species-tiny', "zip", tiny)
```
