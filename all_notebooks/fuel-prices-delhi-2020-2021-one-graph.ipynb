```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (24, 12)
plt.rcParams["font.size"] = 16

f = open("/kaggle/input/fuel-prices-delhi-202021/fuel_prices.csv", "r")

labelx = []

petrol_price = []
diesel_price = []
dollar_to_rupee = []

lines = f.readlines()

for i in range(1, len(lines)):
    line = lines[i].replace("\n", "").split(",")
    labelx.append(line[0])
    petrol_price.append(float(line[1]))
    diesel_price.append(float(line[2]))
    dollar_to_rupee.append(float(line[3]))
    
    

plt.xlabel("2020-2021")
plt.ylabel("Values")
plt.xticks([])
plt.grid(linestyle="--", linewidth=0.618)
plt.plot(labelx, petrol_price, label="petrol_price", linewidth=0.618, marker=".")
plt.plot(labelx, diesel_price, label="diesel_price", linewidth=0.618, marker=".")
plt.plot(labelx, dollar_to_rupee, label="dollar_to_rupee", linewidth=0.618, marker=".")
plt.legend()
plt.show()
```
