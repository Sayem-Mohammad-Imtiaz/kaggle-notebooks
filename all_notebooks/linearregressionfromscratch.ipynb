```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv('/kaggle/input/headbrain/headbrain.csv')
```

```python
data.head()
```

# We will see the Linear Relationship between the Head Size and the Brain Weight

Collecting X and Y

```python
X = data["Head Size(cm^3)"].values
Y = data["Brain Weight(grams)"].values
X.shape
Y.shape
```

Mean X and Mean Y



```python
mean_x = np.mean(X)
print(mean_x)
```

```python
mean_y = np.mean(Y)
```

```python
m = len(X)
```

```python
print(m)
```

# Now calculate b1 and b0 where b1 corresponds to m and b0 corresponds to c

```python
numer = 0
denom = 0
for i in range(m):
    numer += (X[i]-mean_x)*(Y[i]-mean_y)
    denom += (X[i]-mean_x)**2
    b1 = numer / denom
    b0 =  mean_y - (b1*mean_x)
```

```python
print(b1, b0)
```

# Plotting Values and Regression Line

```python
max_x = np.max(X)+100
```

```python
min_x = np.min(X)-100
```

Calculating Values of X and Y

```python
x = np.linspace(min_x, max_x, 1000)
y = b0+b1*x
print(y.shape)
```

Plotting Line

```python
plt.plot(x, y, color = '#58b970', label = 'Regression Line')
#Plotting Scater Points
plt.scatter(X, Y, c = '#ef5423', label = 'Scatter Plot')
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show
```

```python
ss_t = 0
ss_r = 0
for i in range(m):
    y_pred = b0+b1*X[i]
    ss_t += (Y[i]-mean_y)**2
    ss_r += (Y[i]- y_pred)**2
    r2 = 1-(ss_r/ss_t)
  
```
