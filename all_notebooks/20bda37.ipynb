```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
x, y=(df.drop(['UltimateIncurredClaimCost'],axis = 1), df.UltimateIncurredClaimCost)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =
0.3, random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
x.head()
```

```python
x.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
def cross_val(model):
    pred = cross_val_score(model, x, y, cv=10)
    return pred.mean()
```

```python
def print_evaluate(true, predicted):
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R2 Square', r2_square)
    print('__________________________________')
```

```python
def evaluate(true, predicted):
    mae = metrics.mean_absolute_error(true, predicted)
    mse = metrics.mean_squared_error(true, predicted)
    rmse = np.sqrt(metrics.mean_squared_error(true, predicted))
    r2_square = metrics.r2_score(true, predicted)
    return mae, mse, rmse, r2_square
```

```python
x
```

```python
x.isnull().sum()
```

```python
lin_reg = LinearRegression(normalize=True)
lin_reg.fit(x_train,y_train)
```

```python
pred = lin_reg.predict(x_test)
```

```python
test_pred = lin_reg.predict(x_test)
train_pred = lin_reg.predict(x_train)
```

```python
print('Test set evalation:')
print_evaluate(y_test, test_pred)
print('Train set evaluation:')
print_evaluate(y_train, train_pred)
```

# Gradient boost
 

```python
from sklearn.datasets import make_regression
from sklearn.ensemble import GradientBoostingRegressor
gradient=GradientBoostingRegressor(random_state=0)
gradient.fit(x_train,y_train)
```

```python
test_pred2 = gradient.predict(x_test)
train_pred2 = gradient.predict(x_train)
```

```python
print('Test set evalation:')
print_evaluate(y_test, test_pred2)
print('Train set evaluation:')
print_evaluate(y_train, train_pred2)
```

# Randomforest regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression

regr = RandomForestRegressor( random_state=0)
regr.fit(x_train,y_train)

```

```python
test_pred3 = regr.predict(x_test)
train_pred3 = regr.predict(x_train)
```

```python
print('Test set evalation:')
print_evaluate(y_test, test_pred3)
print('Train set evaluation:')
print_evaluate(y_train, train_pred3)
```
