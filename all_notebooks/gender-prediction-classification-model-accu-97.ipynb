```python
import pandas as pd
df = pd.read_csv('/kaggle/input/gender-classification-dataset/gender_classification_v7.csv')
```

```python
df.info()
```

Simply changing the type Dtype of the gender column

```python
twogender = {'Female':0, 'Male':1}
df['gender'] = df['gender'].map(twogender)
```

```python
df.head()
```

```python
X = df.drop('gender', axis=1)
y = df['gender']
```

Split dataset

```python
from sklearn.model_selection import train_test_split, GridSearchCV

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.3)
```

```python
print(X_train.shape)
print(X_val.shape)
print(y_train.shape)
print(y_val.shape)
```

Time to evaluate. I'm using 'accuracy_score' for calculating each algorithms accuracy. I'm also using 'plot_confusion_matrix' for visualising each results

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, plot_confusion_matrix
from sklearn.ensemble import RandomForestClassifier
```

```python
dt = DecisionTreeClassifier(random_state=0)

dt.fit(X_train, y_train)
dt_pred = dt.predict(X_val)
dt_acc = accuracy_score(y_val, dt_pred)
print('Accuracy of Decision Tree is: {:.2f}%'.format(dt_acc*100))
```

```python
plot_confusion_matrix(dt, X_val, y_val)
```

```python
rf = RandomForestClassifier(random_state=0)

rf.fit(X_train, y_train)
rf_pred = rf.predict(X_val)
rf_acc = accuracy_score(y_val, rf_pred)
print('Accuracy of Random Forest is: {:.2f}%'.format(rf_acc*100))
```

```python
plot_confusion_matrix(rf, X_val, y_val)
```

```python
lr = LogisticRegression(random_state=0)

lr.fit(X_train, y_train)
lr_pred = lr.predict(X_val)
lr_acc = accuracy_score(y_val, lr_pred)
print('Accuracy of Logistic Regression is: {:.2f}%'.format(lr_acc*100))
```

```python
plot_confusion_matrix(lr, X_val, y_val)
```

```python
knn = KNeighborsClassifier()
params = {'n_neighbors':[2,3,4,5,6,7,8,9]}

model = GridSearchCV(knn, params, cv=5)
model.fit(X_train, y_train)
model.best_params_
```

```python
kn = KNeighborsClassifier(n_neighbors=8)

kn.fit(X_train, y_train)
kn_pred = kn.predict(X_val)
kn_acc = accuracy_score(y_val, kn_pred)
print('Accuracy of KNeighbors is: {:.2f}%'.format(kn_acc*100))
```

```python
plot_confusion_matrix(kn, X_val, y_val)
```
