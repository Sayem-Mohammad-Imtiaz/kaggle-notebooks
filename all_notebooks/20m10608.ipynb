# Analysis
From here, I will start analyze the data loaded as dataframe object.
I selected ****France**** for this analysis

```python
selected_country = 'France'
df1 = df1[df1['country'] == selected_country]
df2 = df2[df2['country'] == selected_country]
```

I show time-series and color table about Daily COVID-19 cases and Daily vaccinations

```python
#improved visualization
from plotly.offline import iplot
import plotly.graph_objs as go

daily_new_cases = go.Scatter(x=df1.index,y=df1['daily_new_cases'], yaxis='y1', name='Daily confirmed cases')
daily_vaccinations = go.Scatter(x=df2.index,y=df2['daily_vaccinations'], yaxis='y2', name='Daily vaccinations')

Layout_obj = go.Layout(title='COVID vs vaccines', xaxis=dict(title="Date"), yaxis=dict(title='Daily cases'), yaxis2=dict(title='vaccinations', side='right', overlaying='y'))
fig = go.Figure(data=[daily_new_cases, daily_vaccinations], layout=Layout_obj)
iplot(fig)
#fig.write_html('outcome.html')
```

```python
style_object = df1.style.background_gradient(cmap='jet').highlight_max('daily_new_cases').set_caption('Daily cases')
style_object2 = df2.style.background_gradient(cmap='jet').highlight_max('daily_vaccinations').set_caption('Daily vaccinations')
display(style_object, style_object2)
```

I show a global ranking of daily COVID-19 cases and Daily vaccinations per population.
selected country(France) is World No.3 and total vaccinations is World No.56

```python
df1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv', header=0)
df2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv', header=0)

#pop_France = 67012883 
df1 = (df1.groupby('country').sum()['daily_new_cases']).sort_values(ascending=False)
df2 = df2.groupby('country').max()['people_fully_vaccinated_per_hundred'].sort_values(ascending=False)

print('Total COVID-19 cases: ' + str(selected_country) + ' is No.' + str(df1.index.get_loc(selected_country)+1))
print('Total vaccines per number of people: ' + str(selected_country) + ' is No.' + str(df2.index.get_loc(selected_country)+1))

print('----------')
print(df1.to_string())
print('----------')
print(df2.to_string())
```
