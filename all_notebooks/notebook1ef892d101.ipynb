```python
Y = data['posttest']
X = data.drop(['posttest'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3,random_state = 42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
params_options = {'alpha':[0,0.01,0.1,0.3,0.7,1,1.5,2,3,4,5,6,7,8]}
gd_model = GridSearchCV(Lasso(),param_grid = params_options,cv = 5,scoring = 'neg_mean_absolute_error')
gd_model.fit(X_train,Y_train)
gd_model.best_score_
```

```python
y_pred= gd_model.predict(X_test)
```
