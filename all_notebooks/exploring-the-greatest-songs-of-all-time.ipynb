# Top 500 Songs Analysis

## Importing required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Reading the data and quality check

```python
data=pd.read_csv('../input/500-greatest-songs-of-all-time/Top 500 Songs.csv',encoding='unicode_escape')
```

```python
data.head()
```

```python
data.isnull().sum()
```

Streak and position have null values.All other columns have no null columns.

Lets check for duplicates

```python
data.loc[data['title']=='She Loves You']
```

The rows appear to have been duplicated.Lets removed them.

```python
data_clean=data.drop_duplicates()
```

```python
data.shape,data_clean.shape
```

There are 50 duplicates which got removed.

```python
data_clean['writers']=data_clean['writers'].str.strip() ## strip whitespace
```

## Exploratory Data Analysis

```python
data['artist'].nunique()
```

248 artists have composed top 500 greatest songs of all time.Who are the top 5 artists ?

```python
top_artists=data_clean['artist'].value_counts()[:5].reset_index().rename(columns={'artist':'title_count','index':'artist'})
```

```python
top_artists
```

```python
plt.figure(figsize=(8,8))
p=sns.barplot(x='artist',y='title_count',data=top_artists)
plt.title('Greatest artists of all time !!!',fontsize=15)
plt.xlabel('Artist',fontsize=8)
for t in p.patches:
        #print(t)
        p.annotate("{:.0f}".format((t.get_height())), (t.get_x() + t.get_width() / 2., t.get_height()),
             ha='center', va='center', fontsize=15, color='black', xytext=(0, 10),
             textcoords='offset points')
p.set_xticklabels(p.get_xticklabels(),rotation=90)
plt.ylabel('Title Count',fontsize=8)
plt.show()
```

Beatles are in the top spot with 20 labels followed by Bob dylan and Elvis Presley.What are those albums?

```python
data_clean.loc[data_clean['artist']=='The Beatles','title']
```

Who has written these songs ?

```python
data_clean.loc[data_clean['artist']=='The Beatles',['title','writers']]
```

John Lennon and Paul McCartney have written 19 out of 20 songs whereas George Harrison has written one song.Can we check to see whether the songs featured in Greatest songs have been written by same writers for artist or they are different ?

```python
data_clean.loc[data_clean['artist'].isin(top_artists['artist']),['title','writers','artist']].groupby(['artist','writers'])['title'].nunique()
```

Overall,John Lennon and Paul McCartney from The Beatles have produced 19 titles while Mick Jagger,Keith Richards from The rolling stones have 10 titles to their credit.

The hypothesis that the songs are written by same writers for the artist does not hold true.Only for The Beatles it is true.
