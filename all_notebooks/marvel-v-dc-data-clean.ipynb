```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
marvel_dc = pd.read_csv('../input/marvel-vs-dc-imdb-dataset/Marvel_DC_imdb.csv')
```

```python
marvel_dc.info()
```

```python
marvel_dc.head(2)
```

```python
#data cleaning 
#drop unnamed col not needed
marvel_dc = marvel_dc.drop('Unnamed: 0',axis=1)
```

```python
marvel_dc.columns
```

```python
#checking for null
sns.heatmap(marvel_dc.isnull(),cmap='viridis')
```

```python
#Metascore and USA_Gross have a lot of nulls, we can drop
marvel_dc = marvel_dc.drop(['Metascore','USA_Gross'],axis=1)
```

```python
#To answer the question which marvel movie is better, we can drop some more cols that we dont need 
marvel_dc = marvel_dc.drop(['RunTime','Director','Actor'],axis=1)
```

```python
marvel_dc.isnull().sum()
```

```python
#remaining missing values are under 20%, can remove
marvel_dc = marvel_dc.dropna()
```
