```python
import os
import numpy as np
import pandas as pd
import random
import folium
from folium import plugins
import seaborn as sns
import cv2
import matplotlib.pyplot as plt
import IPython
import time
import plotly.express as px
```

```python
data0=pd.read_csv('../input/earthquake-in-japan/japan_2019-2021_query.csv')
data0
```

```python
print(data0.columns.tolist())
```

```python
data0=data0.fillna('N')
data1=data0[['time','latitude','longitude','depth','mag','type']]
data1[0:5]
```

## Year
* **year**=2021

```python
data2=pd.DataFrame(columns=data1.columns.tolist())
for i,item in enumerate(data1['time']):
    if item[0:4]=='2021':
        data2=pd.concat([data2,data1.loc[i:i]])
data2=data2.sort_values(['time']).reset_index(drop=True)        
```

```python
from datetime import datetime as dt
data2a=data2.copy()
date=[]
for item in data2['time']:
    date+=[dt.strptime(item[0:-5],'%Y-%m-%dT%H:%M:%S')]
data2a['date']=date
data2a
```

## Region focused
* 140 < **Longitude** < 144 
* 22 < **Latitude** < 26

```python
data4=data2a[data2a['longitude']<144][data2a['longitude']>140][data2a['latitude']>22][data2a['latitude']<26]
data4=data4.reset_index(drop=True)
data4
```

### All earthquakes occurred 50 km north of the volcano location.
* The volcano location is at LAT 24.284667, LONG 141.481000. 

```python
import plotly.express as px
fig=px.scatter(x=data4['date'],y=data4['mag'])
fig=px.scatter(data4, x="date", y="mag")
fig.update_layout(title='Magnituide of earthquake around Fukutoku-Okanoba region in 2021')
fig.show()
```

```python
data5=data4[['latitude','longitude',]]
for i in range(len(data4)):
    item=data4.loc[i:i,'time'].tolist()[0][0:-5]
    eq_map = folium.Map(location=[24.284667,141.481000],tiles='Stamen Terrain',zoom_start=4.0,min_zoom=2.0)
    eq_map.add_child(plugins.HeatMap(data5.loc[i:i]))
    eq_map.save("./"+str(item)+".html")
```
