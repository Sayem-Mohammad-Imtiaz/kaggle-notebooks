```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
```

```python
data = pd.read_csv("../input/diamonds/diamonds.csv")
data.head()
```

```python
data.drop(data.columns[0], axis=1, inplace=True)
```

```python
data.head()
```

```python
data.shape
```

```python
data["cut"].unique()
data["color"].unique()
data["clarity"].unique()
```

```python
y = data['price']
X = data.drop('price', axis=1)
```

```python
pd.DataFrame(X)
```

```python
encoder = LabelEncoder()

data["cut"] = encoder.fit_transform(data['cut'])
cut_mapping = {index : label for index, label in enumerate(encoder.classes_)}
cut_mapping
```

```python
data["color"] = encoder.fit_transform(data['color'])
color_mapping = {index : label for index, label in enumerate(encoder.classes_)}
color_mapping
```

```python
data["clarity"] = encoder.fit_transform(data['clarity'])
clarity_mapping = {index : label for index, label in enumerate(encoder.classes_)}
clarity_mapping
```

```python
data.head()
```

```python
y = data['price']
X = data.drop('price', axis=1)
```

```python
X.shape
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
std_model = LinearRegression()
l1_model = Ridge(alpha=1)
l2_model = Lasso(alpha=1)

std_model.fit(X_train, y_train)
l1_model.fit(X_train, y_train)
l2_model.fit(X_train, y_train)
```

```python
print(f"Without regression: {std_model.score(X_train, y_train)}")
print(f"Ridge Regression : {l1_model.score(X_train, y_train)}")
print(f"Lasso regression: {l2_model.score(X_train, y_train)}")
```
