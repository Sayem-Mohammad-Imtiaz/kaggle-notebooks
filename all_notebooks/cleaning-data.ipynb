```python
df1 = pd.read_csv('../input/suicide-watch/pushshift_SW_2L.csv')
df2 = pd.read_csv('../input/suicide-watch/pushshift_dep_2L.csv')
```

```python
data = pd.concat([df1, df2])
data.shape
```

```python
data['Post_id'].unique().shape
```

```python
data.isnull().sum()
```

```python
data['Body'].value_counts()
```

```python
data.dropna(axis = 0, how ='any', inplace = True)
data.drop(data[data['Body'] == "[removed]"].index, inplace = True)
data.drop(data[data['Body'] == "[deleted]"].index, inplace = True)
data.drop(data[data['Body'] == "."].index, inplace = True)
data.drop(data[data['Body'] == "..."].index, inplace = True)
data.drop(data[data['Body'] == ":("].index, inplace = True)
data.drop(data[data['Body'] == "Title"].index, inplace = True)
data.shape
```

```python
data['Body'].value_counts()
```

```python
data.drop(data[data['Body'] == "title"].index, inplace = True)
data.drop(data[data['Body'] == "Please"].index, inplace = True)
data.drop(data[data['Body'] == "please"].index, inplace = True)
data.drop(data[data['Body'] == "-"].index, inplace = True)
data.drop(data[data['Body'] == "\n"].index, inplace = True)
data.shape
```

```python
data['Subreddit'].value_counts()
```

```python
data.shape
```

```python
data.info()
```

```python
data = data.sample(frac=1).reset_index(drop=True)
data.head(5)
```
