Create data frame

Changing dates to numerical values

```python
dataset = pd.read_csv('/kaggle/input/covid19-confirmed-cases-kerala/Confirmed Covid 19 Cases Kerala.csv')

X = [] # Create X as a list
for i in range(0, len(dataset['Date'])): #Loop through the date column 
  X.append(i) # Append the number i as the number of days
X = np.asarray(X) # Change X to numpy array
X = X.reshape(-1,1) # Reshape X 
y = dataset.iloc[:,-1].values
```

Splitting

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

Training using Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 15, random_state = 0)
regressor.fit(X, y)
```

Traning using polynomial regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 6)
X_poly = poly_reg.fit_transform(X)
lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
```

Visualizing Random Forest Regression

```python
from matplotlib.pyplot import figure
X_grid = np.arange(min(X), max(X), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
pred = regressor.predict(X_grid)

figure(figsize=(20, 10), dpi=80)

fig, ax = plt.subplots(2,1 ,figsize=(15,15))

ax[0].plot(X, y, color = 'tomato', label = 'Real')
ax[0].plot(X_grid, pred, color = 'navy' , label = 'Prediction')
ax[0].set_title('Covid-19 Cases')
ax[0].set_xlabel('Days')
ax[0].set_ylabel('Cases')
ax[0].legend(loc = 'upper left')


ax[1].scatter(X, y, color = 'red', label = 'Real')
ax[1].plot(X_grid, pred, color = 'navy', label = 'Prediction')
ax[1].set_title('Covid-19 Cases')
ax[1].set_xlabel('Days')
ax[1].set_ylabel('Cases')
ax[1].legend(loc = 'upper left')

plt.show()
```

Visualizing Polynomial Regression

```python
from matplotlib.pyplot import figure

figure(figsize=(13, 8), dpi=80)

X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'tomato', label = 'Real')
plt.plot(X_grid, lin_reg.predict(poly_reg.fit_transform(X_grid)), color = 'navy', label = 'Prediction')
plt.title('Covid-19 Cases')
plt.xlabel('Days')
plt.ylabel('Cases')
plt.legend(loc = 'upper left')
plt.show()
```
