### Item B - Tabela de frequência

Construa uma tabela de frequência para cada uma das **variáveis qualitativas** que você escolheu (caso não tenha escolhido nenhuma, deixe esta questão em branco). Uma dica: a função *value_counts()* do Pandas pode ser muito útil. =)


```python
specific_columns = ['aeronave_operador_categoria','aeronave_tipo_veiculo','aeronave_motor_tipo','aeronave_motor_quantidade','aeronave_pais_fabricante','aeronave_nivel_dano']

freq_dfs = {}

for column in specific_columns:
    print("Tabela de frequência para a variável "+column)
    freq_dfs[column] = df_original[column].value_counts().to_frame().reset_index(inplace=False).rename(columns = {''+column:'qtde'}).rename(columns = {'index':column})
    display(freq_dfs[column].style.hide_index())
    print("____")
```

### variável aeronave_tipo_veiculo, gráfico de barra

```python
def plot_horizontal_bar(df,subplot_index,title):
    df = df.sort_values('qtde',ascending=1)
    
    ax = plt.subplot(1,2,subplot_index)
    
    if subplot_index==1:
        plt.ylabel(varname)
        color = ['powderblue']+['yellowgreen' for i in range(len(df)-1)]
    else:
        color = 'powderblue'
    
    ax.barh(df[varname], df['qtde'],   label = varname, color = color)
    plt.xticks(rotation='vertical')

    
    plt.xlabel("Quantidade")
    ax.set_title(title)
    plt.legend()
    for i in range(len(df['qtde'])):
        ax.annotate(
            df['qtde'].values[i],
            xy=(df['qtde'].values[i],i),
            horizontalalignment="center",
            verticalalignment="center",
            weight='bold',
            bbox=dict(facecolor='white', edgecolor='none', pad=2),
            xytext=(40,0),textcoords='offset pixels',arrowprops=dict(facecolor='black', shrink=0.05, width = 0.8,headwidth=5)
        )

varname = 'aeronave_tipo_veiculo'
df = freq_dfs['aeronave_tipo_veiculo']

df_principais = df[(df[varname]=="AVIÃO")|(df[varname]=="HELICÓPTERO")|(df[varname]=="ULTRALEVE")]
df_outros = df[(df[varname]!="AVIÃO")&(df[varname]!="HELICÓPTERO")&(df[varname]!="ULTRALEVE")]
total_outros = df_outros['qtde'].sum()
new_row = {}
new_row[varname] = 'OUTROS'
new_row['qtde'] = total_outros
df_principais = df_principais.append(new_row, ignore_index=True)

fig=plt.figure(figsize=(20,10))
fig.suptitle("Tipo da aeronave vs Quantidade de ocorrências")
plot_horizontal_bar(df_principais,1,"Principais tipos")
plot_horizontal_bar(df_outros,2,"Breakdown de outros")

plt.show()
```

```python
def print_polar(df,subplot_index,title=None):
    # Compute pie slices
    N = len(df)

    theta_step = 2 * np.pi / 8
    theta = [np.pi/8 + theta_step*i for i in range(N)]

    radii = df['qtde'].values
    width = np.pi/4

    colors = plt.cm.plasma(radii / 5000.)
    
    ax = plt.subplot(1,2,subplot_index, projection='polar')
    ax.set_title(title)
    ax.bar(theta, radii, width=width, bottom=1, color=colors, alpha=0.5)
    ax.set_rlabel_position(0)
    plt.yticks(fontsize=11,weight="bold")
    i = 0
    for row in df.itertuples():
        ax.annotate(getattr(row,varname)+", "+str(row.qtde),
                    xy=(theta[i], df['qtde'].max()),  # theta, radius
                    horizontalalignment='center',
                    verticalalignment='bottom',
                    weight='bold',
                    bbox=dict(facecolor='white', edgecolor='none', pad=2),
                    )
        i=i+1
        
import numpy as np
varname = 'aeronave_motor_tipo'

df = freq_dfs[varname]

## total do "Outros"
df_outros = df[(df[varname]!="PISTÃO") & (df[varname]!="JATO")]
total_outros = df_outros['qtde'].sum()

## fazendo dos motores principais:
df_resumo = df[(df[varname]=="PISTÃO") | (df[varname]=="JATO")]
new_row = {}
new_row[varname] = "OUTROS"
new_row['qtde'] = total_outros
df_resumo = df_resumo.append(new_row,ignore_index=True)


fig=plt.figure(figsize=(16,8))
fig.suptitle("Tipo de motor vs número de ocorrências", fontsize = 16)
print_polar(df_resumo,1,"Principais tipos")
print_polar(df_outros,2,"Breakdown do Outros")


plt.show()
```

```python
import geopandas

countries_df = freq_dfs['aeronave_pais_fabricante']
countries_df['iso'] = countries_df['aeronave_pais_fabricante']
countries_df['iso'].replace({"ESTADOS UNIDOS":"USA","BRASIL":"BRA","FRANÇA":"FRA","NÃO IDENTIFICADO":"ND","ARGENTINA":"ARG","HOLANDA":"NLD","CANADÁ":"CAN","ITÁLIA":"ITA","JAPÃO":"JPN","AUSTRIA":"AUT","REPÚBLICA TCHECA":"CZE","ALEMANHA":"DEU","SUIÇA":"CHE","INGLATERRA":"GBR","POLÔNIA":"POL","UCRÂNIA":"UKR"},inplace=True)

world = geopandas.read_file(geopandas.datasets.get_path('naturalearth_lowres'))


## adicionando dados de aeronaves fabricadas
world['aeronaves_fabricadas'] = [0 for i in range(len(world))]
def func(x):
    if not countries_df[countries_df['iso'] == x['iso_a3']].empty:

        x['aeronaves_fabricadas'] = countries_df[countries_df['iso'] == x['iso_a3']].values[0][1]
    return x
world = world.apply(func,axis=1 )

## criando coluna de coordenadas xy para anotações no mapa
world['coords'] = world.apply(lambda x : x['geometry'].representative_point().coords[0],axis=1)

## criando plots básicos do mapa mundi
base = world.plot(figsize=(32, 16), vmin=1, color ='gray', edgecolor="black")
world_filtered = world[world['aeronaves_fabricadas']!=0]
world_filtered.plot(ax=base,figsize=(32, 16), vmin=1, column = 'aeronaves_fabricadas', cmap='OrRd',edgecolor="black", legend = True)
plt.title("Número de aeronaves fabricadas por país, dentro do registro de acidentes aéreos", fontsize=18)

europe_total = 0
for row in world_filtered.itertuples():
    if not row.continent == "Europe":
        plt.annotate(s=row.name+", "+str(row.aeronaves_fabricadas),xy=row.coords,weight='bold', fontsize=18,xytext=(100,5),textcoords='offset pixels',arrowprops=dict(facecolor='black', shrink=0.05), bbox=dict(facecolor='white', edgecolor='none', pad=2),)
    else:
        europe_total += row.aeronaves_fabricadas
plt.annotate(s="Europe, "+str(europe_total),xy=world[world['iso_a3']=="FRA"]['coords'].values[0],weight='bold', fontsize=18,xytext=(100,15),textcoords='offset pixels',arrowprops=dict(facecolor='black', shrink=0.05), bbox=dict(facecolor='white', edgecolor='none', pad=2),)
    
## criando plot zoom na europa

## limpando território francês
world.loc[world['iso_a3']=="FRA",'geometry'] = world[world['iso_a3']=="FRA"]['geometry'].values[0][2]
## limpando território norueguês
world.loc[world['iso_a3']=="NOR",'geometry'] = world[world['iso_a3']=="NOR"]['geometry'].values[0][0]

europe = world[(world['continent']=="Europe") & (world['name']!="Russia")]
europe_base = europe.plot(figsize=(32, 16), vmin=1, color ='white', edgecolor="black")
europe_filtered = europe[europe['aeronaves_fabricadas']!=0]
europe_filtered.plot(ax=europe_base,figsize=(32, 16), vmin=1, column = 'aeronaves_fabricadas',cmap='OrRd', edgecolor="black", legend = True)
plt.title("Número de aeronaves fabricadas por país na Europa, dentro do registro de acidentes aéreos", fontsize=18)

for row in europe_filtered.itertuples():
    plt.annotate(s=row.name+", "+str(row.aeronaves_fabricadas),xy=row.coords,weight='bold', fontsize=14, bbox=dict(facecolor='white', edgecolor='none', pad=2),)
plt.show()
```
