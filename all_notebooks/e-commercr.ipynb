```python
#统计每个顾客最近一次消费时间、消费次数，购买数量和消费总金额

rfm = df1[df1['Quantity'] > 0].groupby('CustomerID').agg({'InvoiceDate':np.max,'InvoiceNo':'nunique','Price':np.sum})
```

```python
#计算最后一次消费时间距离现在有多远

rfm['R'] = -(rfm.InvoiceDate - rfm.InvoiceDate.max()) / np.timedelta64(1,'D')
```

```python
#对列名进行重命名

rfm.rename(columns={'InvoiceNo':'F','Price':'M'},inplace=True)
```

```python
rfm.head()
```

```python
def rfm_func(x):
    level = x.apply(lambda x:'1' if x >= 0 else '0')
    label = level.R + level.F + level.M
    d = {
        '111':'重要价值客户',
        '011':'重要保持客户',
        '101':'重要挽留客户',
        '001':'重要发展客户',
        '110':'一般价值客户',
        '010':'一般保持客户',
        '100':'一般挽留客户',
        '000':'一般发展客户'
    }
    result = d[label]
    return result   
```

```python
rfm['label'] = rfm[['R','F','M']].apply(lambda x : x - x.mean()).apply(rfm_func,axis = 1)
```
