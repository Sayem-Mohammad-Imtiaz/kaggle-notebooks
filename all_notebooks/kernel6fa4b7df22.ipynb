## Desafio
Como o desafio é analisar um atributo desse problema, como exemplo iremos trabalhar com o atributo pclass, que segundo a descrição desse dado (https://www.kaggle.com/c/titanic/data) ele representa a classe da cabine em que uma pessoa estava hospedada, sendo que existem somente 3 classes: 1ª, 2ª ou 3ª.

O objetivo é responder as seguintes perguntas:

A) Esse atributo é importante para a resolução do problema? Por que?

B) Existem dados faltantes desse atributo? Quantos? Como podemos tratá-los e por que fazer dessa
forma?

C) O dado possui algum tipo de erro? (Ex: digitação incorreta, etc)

Vamos começar respondendo a pergunta C. 
### O atributo pclass possui algum tipo de erro?

Pelo comando head() do pandas conseguimos visualizar os valores de cada amostra, uma maneira ingênua de realizar essa análise seria olhar cada amostra procurando se existe algum valor errado. Se fizermos dessa forma, quantas amostras teremos que analisar?

```python
print("A base de dados possui o total de %d linhas e %d colunas" % train_data.shape)
```

O comando shape retorna as dimensões do nosso dataset, lembrando que o dataset nada mais é que uma matriz, conforme estudamos no capítulo 2.1. Dessa forma, a primeira dimensão representa o número de linhas e a segunda representa o número de colunas

Sendo assim, teríamos que analisar 891 amostras para dizer se o atributo Pclass possui algum dado errado, um tanto ineficiente, concordam?

Para nos ajudar com essa tarefa, utilizaremos o método unique() do pandas e numpy que retornam os valores únicos de um atributo. Ou seja, só existem amostras com os valores que o método unique() retornar

```python
train_data.Pclass.unique()
```

Como os únicos valores retornados foram (1, 2, 3) podemos ter certeza de que não existem dados errados, pois conforme discutimos anteriormente 1, 2, 3 são os únicos valores possíveis para esse atributo.

### Existem dados faltantes desse atributo? Quantos? Como podemos tratá-los e por que fazer dessa forma?

Para responder essa pergunta, iremos utilizar o método info() do pandas, que mostra a quantidade de valores não nulos para cada atributo

```python
train_data.info()
```

Podemos ver que o atributo Pclass possui 891 valores não nulos, e como a quantidade total de amostras também é 891, logo vemos que não existem valores nulos.

### Esse atributo é importante para a resolução do problema? Por que?

Para responder essa pergunta, vamos analisar quantas pessoas sobreviveram e quantas morreram por classe?

```python
train_data.hist(column="Pclass")
```

O método hist() do pandas mostra o histograma de um determinado atributo (definido pelo parâmetro column), porém não dividide em sobreviventes e não sobreviventes, para isso passaremos mais um parâmetro chamado ` by ` que permite agrupar os dados por outro atributo.

Lembrando que o histograma mostra a frequência de cada valor para um determinado atributo.

```python
train_data.hist(column="Pclass", by="Survived")
```

No primeiro gráfico é apresentado o histograma das pessoas que não sobreviveram, sendo que a terceira classe possui o maior número de mortes.

O cenário é invertido no segundo gráfico, onde a maior parte da população que sobreviveu pertencia a primeira classe.

Com esse gráfico conseguimos justificar que o atributo Pclass** é importante** para a resolução do problema, além de **não possuir dados faltantes** e **não possuir erros**.
