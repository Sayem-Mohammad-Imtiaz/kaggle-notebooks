```python
import numpy as np
import pandas as pd
import sklearn as sk
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("../input/football-player/players_20_numeric_normalized.csv")
X = StandardScaler().fit_transform(df)
```

```python
db = DBSCAN(eps=5, min_samples=50).fit(X)
```

```python
print(len(set(db.labels_)) - (1 if -1 in db.labels_ else 0))
print(sk.metrics.calinski_harabasz_score(X,db.labels_))
print(sk.metrics.davies_bouldin_score(X,db.labels_))
#print(sk.metrics.silhouette_score(X,db.labels_))
```

```python
df_o = df[df['gk_speed']==0]
# print(df_gk)

df_gk = df[df['gk_speed']!=0]
NUM_GOALKEEPER = df_gk.shape[0]
# print(df_o)
print(NUM_GOALKEEPER)

db = DBSCAN(eps=5, min_samples=50).fit(StandardScaler().fit_transform(df_o))
print(len(set(db.labels_)) - (1 if -1 in db.labels_ else 0))
print(sk.metrics.calinski_harabasz_score(df_o,db.labels_))
```

```python
label_gk = np.full(shape = 2036, fill_value = 3, dtype = np.int)
label_concat = np.concatenate((label_gk, db.labels_))

df_concat = pd.concat([df_gk, df_o])
print(sk.metrics.calinski_harabasz_score(df_concat, label_concat))
print(sk.metrics.davies_bouldin_score(df_concat, label_concat))
```
