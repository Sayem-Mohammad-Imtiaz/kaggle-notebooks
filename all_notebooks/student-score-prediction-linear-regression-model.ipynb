```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df=pd.read_csv('../input/student-study-hour-v2/Student Study Hour V2.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
g=sns.JointGrid(data=df,x='Hours',y='Scores')
g.plot(sns.regplot,sns.histplot);
```

```python
df['intercept']=1
lm=sm.OLS(df['Scores'],df[['intercept','Hours']])
results=lm.fit()
results.summary()
```

----

y=b0+b1x

- y.... Scores (Response variable that we want to predict)

- x.... Hours  (Explanatory variable that we use as an input to predict student score)

b0=3.2017
b1=9.6774
