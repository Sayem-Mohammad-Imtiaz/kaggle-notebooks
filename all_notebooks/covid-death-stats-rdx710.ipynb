```python
df = pd.read_csv('../input/covid19-deaths-dataset/all_weekly_excess_deaths.csv')
# df.head()
df_temp = df['end_date'].str.split('-', expand=True)[[1,0]]
df['Date'] = df_temp[1] + '/' + df_temp[0]
df.head()
```

## Total World 2020, 2021 comparision

```python
sns.pairplot(df, vars = ['total_deaths', 'covid_deaths', 'non_covid_deaths'], hue = 'year')
```

```python
country = ''
country_list = []

for country_name in df.country:
    if(country_name != country):
        country = country_name
#         print(country_name)
        country_list.append(country_name)
        
country_list
```

## Dataset for year 2020

```python
df_yr2020 = df[df.year == 2020]
df_yr2020
```

### Total Deaths (2020)

```python
fig = px.choropleth(df_yr2020.rename(columns={'total_deaths':'Total Deaths'}), 
                    locations='country',
                    locationmode = 'country names',
                    color='Total Deaths', 
                    animation_frame='end_date'
                   )
fig.update_layout(geo=dict(
                  showframe = False,
                  showcoastlines = False)
                )
fig.show()
```

### Covid Deaths (2020)

```python
fig = px.choropleth(df_yr2020.rename(columns={'covid_deaths':'Covid Deaths'}), 
                    locations='country',
                    locationmode = 'country names',
                    color='Covid Deaths', 
                    animation_frame='end_date'
                   )
fig.update_layout(geo=dict(
                  showframe = False,
                  showcoastlines = False)
                )
fig.show()
```

### Non-Covid Deaths (2020)

```python
fig = px.choropleth(df_yr2020.rename(columns={'non_covid_deaths':'Non-Covid Deaths'}), 
                    locations='country',
                    locationmode = 'country names',
                    color='Non-Covid Deaths', 
                    animation_frame='end_date'
                   )
fig.update_layout(geo=dict(
                  showframe = False,
                  showcoastlines = False)
                )
fig.show()
```

### Graph showing No. of total deaths and covid deaths per month for year 2020

```python
plt.figure(figsize = (15, 10))
sns.barplot(data = df_yr2020, x = 'Date', y = 'total_deaths', color = 'red', label = 'Total Deaths')
sns.barplot(data = df_yr2020, x = 'Date', y = 'covid_deaths', color = 'yellow', label = 'Covid Deaths')
plt.xlabel(xlabel = 'Date', fontsize = 16, fontweight = 'bold')
plt.ylabel(ylabel = 'Number of Deaths', fontsize = 16, fontweight = 'bold')
plt.legend()
plt.show()
```

## Dataset for United States

```python
df_us = pd.read_csv('../input/covid19-deaths-dataset/us-counties.csv')
df_us
```
