 In this lesson I will be working on Outlier detection and removal of the outlier by Using IQR
The Data points consider  outliers if it's below the first quartile or above the third quartile

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
Data=pd.read_csv("/kaggle/input/weight-height/weight-height.csv")
Data.head()
```

```python
#Scatter plot to show the outliers
plt.figure(figsize=(20,8))
plt.scatter(Data.Weight,Data.Height)
plt.xlabel('Weight')
plt.ylabel('Height')
plt.show()
```

```python
Data.describe()
```

```python
#Height Plot Distrbution
plt.figure(figsize=(20,8))

plt.subplot(1,3,1)
sns.distplot(Data.Height)
plt.ylabel('Distrbution')

plt.subplot(1,3,2)
sns.boxplot(y=Data.Height)

plt.subplot(1,3,3)
plt.hist(Data.Height, bins=20, rwidth=0.8)
plt.xlabel('Height')
plt.ylabel('Count')

plt.show()
```

Q1 is the first Quartile of the data and it is about 25% of the data
Q3 is the third quartile of the data and it is about 75% of the Data
Inter-Quartile Range or IQR id the difference between Q3 and Q1

```python
Q1=Data.Height.quantile(0.25)
Q3=Data.Height.quantile(0.75)
Q1,Q3
```

```python
IQR=Q3-Q1
IQR
```

1.5 is the scale that we are taking to limit the outliers

```python
lower_limit=Q1-1.5*IQR
Upper_limit=Q3+1.5*IQR
lower_limit,Upper_limit
```

```python
#Height outliers
Data_outlier=Data[(Data.Height<lower_limit)|(Data.Height>Upper_limit)]
Data_outlier
```

```python
Data_no_outlier=Data[(Data.Height>lower_limit)&(Data.Height<Upper_limit)]
Data_no_outlier
```

```python
#Weight Plot Distrbution
plt.figure(figsize=(20,8))

plt.subplot(1,3,1)
sns.distplot(Data.Weight)
plt.ylabel('Distrbution')

plt.subplot(1,3,2)
sns.boxplot(y=Data.Weight)

plt.subplot(1,3,3)
plt.hist(Data.Weight, bins=20, rwidth=0.8)
plt.xlabel('Weight')
plt.ylabel('Count')
plt.show()
```

```python
Q1=Data.Weight.quantile(0.25)
Q3=Data.Weight.quantile(0.75)
Q1,Q3
```

```python
IQR=Q3-Q1
IQR
```

```python
Weight_lower_limit=Q1-1.5*IQR
Weight_Upper_limit=Q3+1.5*IQR
Weight_lower_limit,Weight_Upper_limit
```

```python
# Weight outliers
Data_outlier=Data[(Data.Weight<Weight_lower_limit)|(Data.Weight>Weight_Upper_limit)]
Data_outlier
```

```python
Data_no_outlier=Data[(Data.Weight>Weight_lower_limit)&(Data.Weight<Weight_Upper_limit)]
Data_no_outlier
```
