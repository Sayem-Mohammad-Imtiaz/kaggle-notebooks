Lets look at distributions

```python
import plotly.express as px
```

```python
## The majority of the content listed was in fact movies rather than TV shows (5377 vs 2410) (i.e. ~ 66%:33%)
fig = px.histogram(df, x="type")
fig.show()
```

```python
## There is a long tail of directors across the different shows.
### Jan Suter, Raul Campos appears together the most
### Other common names are Marcus Raboy, Jay Karas, Cathy Garcia-Molina
### Lets see if we recognise their work?

fig = px.histogram(df, x="director").update_xaxes(categoryorder="total descending")
fig.show()
```

```python
fig = px.histogram(df, x="release_year").update_xaxes(categoryorder="total descending")
fig.show()
```

```python
df['duration_mins'] = df['duration'].apply(lambda x: int(x[0:x.find('min')]) if x.find('min') != -1 else 0)
df['release_year_grouped'] = df['release_year'].apply(lambda x: str(x) if 2016 < x else '<2016')
```

```python
fig = px.histogram(df[df['type']=='Movie'],
                   x="duration_mins")
fig.show()
```

```python
fig = px.histogram(df[df['type']=='Movie'],
                   x="duration_mins",
                   color="release_year_grouped")
fig.show()
```

```python
df['duration_seasons'] = df['duration'].apply(lambda x: int(x[0:x.find('Season')]) if x.find('Season') != -1 else 0)
```

```python
fig = px.histogram(df[df['type']=='TV Show'],
                   x="duration_seasons")
fig.show()
```

```python
fig = px.histogram(df[df['type']=='TV Show'],
                   x="duration_seasons",
                   color="release_year_grouped")
fig.show()
```

```python
fig = px.histogram(df[df['type']=='Movie'],
                   x="rating")
fig.show()
```

```python
fig = px.histogram(df[df['type']=='TV Show'],
                   x="rating")
fig.show()
```

```python
fig = px.histogram(df[df['type']=='Movie'],
                   x="country").update_xaxes(categoryorder="total descending")
fig.show()
```

```python
fig = px.histogram(df[df['type']=='Movie'],
                   x="country",
                   color="release_year_grouped").update_xaxes(categoryorder="total descending")
fig.show()
```
