```python
plt.figure(figsize=(30,10))
sns.countplot(data['country'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=data20,x="country",y="population",marker='o')
plt.title("country and population")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=data20,x="country",y="expected_deaths",marker='o' ,color= 'red')
plt.title("country and expected_deaths")
plt.xticks(rotation=90)
plt.show()
```
