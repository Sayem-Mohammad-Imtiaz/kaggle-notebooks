```python
import pandas as pd
df=pd.read_csv('/kaggle/input/pima-indians-diabetes-database/diabetes.csv')
df
```

```python
X=df.drop(['Outcome'],axis=1)
```

```python
X
```

```python
y=df['Outcome']
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmaxscaler=MinMaxScaler()
X_norm=minmaxscaler.fit_transform(X)
```

```python
df_norm=pd.DataFrame(X_norm,columns=X.columns)
df_norm
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)


X_pca=pca.fit_transform(df_norm)
```

```python
df_pca=pd.DataFrame(X_pca,columns=['coluna_1','coluna_2'])
df_pca
```

```python
df_pca['diabetes']=y
```

```python
import altair as alt
alt.Chart(df_pca).mark_point().encode(
x='coluna_1',
y='coluna_2',
color='diabetes')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_pca,y)

```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import f1_score

for clf in [DecisionTreeClassifier(),SVC(),MLPClassifier(),RandomForestClassifier()]:
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    print(clf,f1_score(y_test,y_pred))
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif


# Create an SelectKBest object to select features with two best ANOVA F-Values
fvalue_selector = SelectKBest(f_classif, k=2)

# Apply the SelectKBest object to the features and target
X_kbest = fvalue_selector.fit_transform(X, y)


from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_kbest,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix

for clf in [DecisionTreeClassifier(),SVC(),MLPClassifier(),RandomForestClassifier()]:
    clf.fit(X_train,y_train)
    y_pred=clf.predict(X_test)
    print(clf,f1_score(y_test,y_pred))
    print(confusion_matrix(y_test,y_pred))
```
