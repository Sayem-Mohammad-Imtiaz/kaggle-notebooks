Import the importance libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, mean_absolute_error
%matplotlib inline
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

Load data from data.csv file

```python
data=pd.read_csv("/kaggle/input/breast-cancer-wisconsin-data/data.csv")
```

Clear data from columns with NAN or not important, then displaying the info of each column

```python
data.dropna( axis = 1, inplace = True )
data.drop( columns = ['id'], inplace = True )
data.info()
```

Split data based on diagnosis column to check unnecessary columns

```python
data_b=data[data['diagnosis']=='B']
data_b.describe()
```

```python
data_m=data[data['diagnosis']=='M']
data_m.describe()
```

```python
fig, ax=plt.subplots(figsize=(10,10))
ax.hist(data_b['area_mean'],alpha=0.5,label='benign')
ax.hist(data_m['area_mean'],alpha=0.5,label='malignant')
ax.set_title('Benign and Malignant destributions of area mean')
ax.set_xlabel('area_mean')
ax.set_ylabel('count')
ax.legend(loc='upper right')
plt.show();
```

Deleted data that unaffected  my analysis and has the same variance, then determined both X and y for modeling 

```python
y=data.diagnosis
X=data.drop(columns = ['diagnosis','fractal_dimension_mean','smoothness_se'])
```

Split data to training and testing parts

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

Optimizing the DecisionTreeClassifier model

```python
accuracy_level=[]
max_depths=range(1,11)
random_states=range(1,11)
max_leafs_nodes=range(2,11)
for max_depth in max_depths:
    for random_state in random_states:
        for max_leaf_nodes in max_leafs_nodes:
            tree_model = DecisionTreeClassifier(max_depth=max_depth, random_state=random_state, max_leaf_nodes=max_leaf_nodes)
            tree_model.fit(X_train,y_train)
            y_pred=tree_model.predict(X_test)
            accuracy_level.append(accuracy_score(y_test,y_pred)*100)
print(max(accuracy_level))            
```
