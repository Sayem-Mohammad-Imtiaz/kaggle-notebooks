### Crie um histograma da distribuição de idades (age)

### Separe o dataset nos atributos de entrada X e objetivo y

```python
dados.keys()
```

```python
X = dados.loc[:,'age':'histology']
y = dados.loc[:,['class']]
```

```python
X.head()
```

```python
y.head()
```

### Utilize a biblioteca de pré-processamento para normalizar os atributos de entrada (X)

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X = ss.fit_transform(X)
```

### Separe o dataset em Treino e Teste

```python

from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
```

### Crie o modelo para MLP, treine-o e faça a predição com os dados de teste (y_pred)

```python
from sklearn.linear_model import Perceptron
pp = Perceptron()
modelo = pp.fit(X_train, y_train)
y_predict = modelo.predict(X_test)
```

### Calcule o Score do Modelo

```python
score = modelo.score(X_test, y_test)
score
```
