# Comparing No of bestsellers between fictional and non fictional genre books

```python
import pandas as pd
import numpy as np
import plotly.offline as pyo
import plotly. graph_objs as go
temp_df_fiction = df[df['Genre']=='Fiction'].value_counts().sum()
temp_df_non_fiction = df[df['Genre']=='Non Fiction'].value_counts().sum()


labels=['Fiction','Non_Fiction']
values = [temp_df_fiction,temp_df_non_fiction]
fig = go.Figure(data=[go.Pie(labels = labels, values=values)])
fig.show()
#df_non_fiction.value_counts().sum()
```

**Non Fictional books are sold more than fictional books**

# Bestsellers w.r.t Author

```python
import pandas as pd
import numpy as np
import plotly.offline as pyo
import plotly. graph_objs as go

#Author vs prices
new_df = df.groupby('Author').mean().reset_index().sort_values('Price', ascending =False).head(10)
data  = [go.Bar(x = new_df['Author'], y = new_df['Price'] )]
layout = go.Layout(template = 'plotly_dark', title = 'Top 10 Bestsellers of Authors with highest prices of books')
fig = go.Figure(data=data, layout = layout)
fig.show()

#Author vs Name
new_df = df.groupby('Author').count().reset_index().sort_values('Name', ascending=False).head(10)
data =[go.Bar(x= new_df['Author'], y=new_df['Name'])]
layout = go.Layout(template = 'plotly_dark', title = 'TOp 10 Authors with maximum no of bestsellers ')
fig = go.Figure(data = data, layout=layout)
fig.show()

#Author vs User Rating
new_df = df.groupby('Author').mean().reset_index().sort_values('Reviews', ascending=False).head(10)
data =[go.Bar(x= new_df['Author'], y=new_df['Reviews'])]
layout = go.Layout(template = 'plotly_dark', title = 'Top 10 Authors with maximum no of bestsellers having highest reviews' )
fig = go.Figure(data = data, layout=layout)
fig.show()

```

# Bestsellers w.r.t Genre

```python
import pandas as pd
import numpy as np
import plotly.offline as pyo
import plotly. graph_objs as go

#Genre vs Price
new_df1 = df[df['Genre'] =='Fiction']
new_df2 = df[df['Genre'] =='Non Fiction']

trace0 = go.Bar(x = new_df1['Year'] ,y = new_df1['Price'], name = 'Fiction' )
trace1 = go.Bar(x = new_df2['Year'] ,y = new_df2['Price'], name = 'Non_Fiction')
data = [trace0,trace1]

layout = go.Layout(title = 'Price by Genre over the years', barmode='group')
fig = go.Figure(data = data , layout = layout)
fig.show()

#Genre vs Reviews
new_df1 = df[df['Genre'] =='Fiction']
new_df2 = df[df['Genre'] =='Non Fiction']

trace0 = go.Bar(x = new_df1['Year'] ,y = new_df1['Reviews'], name = 'Fiction' )
trace1 = go.Bar(x = new_df2['Year'] ,y = new_df2['Reviews'], name = 'Non_Fiction')
data = [trace0,trace1]

layout = go.Layout( title = 'Reviews by Genre over the years', barmode='stack')
fig = go.Figure(data = data , layout = layout)
fig.show()
```

# Bestseller books with maximum ratings and reviews

```python
import pandas as pd
import numpy as np
import plotly.offline as pyo
import plotly. graph_objs as go
new_df = df.groupby('Name').mean().reset_index().sort_values('User Rating', ascending = False).head(10)
x= new_df['Name']
y=new_df['Reviews']
z=np.array(new_df['User Rating'])
trace = go.Heatmap(
   x = x,
   y = y,
   z = z,
   type = 'heatmap',
   colorscale = 'Viridis'
)
layout = go.Layout(title = 'Top 10 books with maximum user rating and reviews')
data = [trace]
fig = go.Figure(data = data, layout = layout)
fig.show()
```
