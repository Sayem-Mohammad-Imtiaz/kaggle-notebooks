```python
# Line Plot
# color = color, label = label,
data.Km.plot(kind = 'line', color = 'g',label = 'Km',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
data.Fiyat.plot(color = 'r',label = 'Fiyat',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Line Plot')
plt.show()
```

```python
# Scatter Plot 
# x = Fiyat, y = Km
data.plot(kind='scatter', x='Fiyat', y='Km',alpha = 0.5,color = 'red')
plt.xlabel('Fiyat')              # label = name of label
plt.ylabel('Km')
plt.title('Fiyat Km Scatter Plot')   
```

```python
# Histogram
# bins = number of bar in figure
data.Km.plot(kind = 'hist',bins = 50,figsize = (12,12))
plt.show()
```

```python
# clf() = cleans it up again you can start a fresh
data.Fiyat.plot(kind = 'hist',bins = 50)
plt.clf()
# We cannot see plot due to clf()
```
