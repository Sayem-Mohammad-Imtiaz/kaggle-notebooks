Önce sadece ihtiyacımız olan parametreler kalması için data setini kırpalım. #Bunu sadece pratik olması için yapıyorum tahminde en iyi sonuç için "state"le de işlemler yapılabilir. 

```python
df=df[["R&D Spend","Administration","Marketing Spend","Profit"]]
```

Kırpılmış halini görmek için tekrar ilk 5 elemanı yazdıralım.

```python
df.head()
```

Tahmin edeceğimiz özelliği belirtelim

```python
predict = "Profit"

X = np.array(df.drop([predict], 1)) 
y = np.array(df[predict]) 
```


Data seti *train* ve *test* olarak ikiye ayıralım


```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
```

Test edilecek değerleri ayırmalıyız ki sonuç mantıklı çıksın. Ayırmadığımız halinde hep yüksek doğrulukla tahmin etmiş olacağız.

```python
x_test
```

```python
y_test
```

Kullanacağımız modeli tanımlayalım, metrics'i çekirdeğe dahil edelim

```python
from sklearn.linear_model import LinearRegression

linear = LinearRegression()
```

Dataseti trainleyelim

```python
linear.fit(x_train, y_train)
acc = linear.score(x_test, y_test) #acc = accuracy
```


Train sonucunu yazdıralım


```python
print(acc)
```

linear regresyon doğrusunun ekseni kestiği nokta ve eğimini yazdıralım

```python
print('Coefficient: \n', linear.coef_ ) # These are each slope value
print('Intercept: \n', linear.intercept_ ) # This is the intercept
```

Tahmin sonuçlarını y_pred değişkenine atayalım ve yazdıralım.

```python
y_pred = linear.predict(x_test)
y_pred
```

Tahminlerimizin yakınlığını tek tek inceleyeceğimiz tabloya yazdıralım.

```python
test = pd.DataFrame(x_test).copy() # copy kullanılmazsa değişikler x_test için de gerçekleşir.
test["Profit"] = y_test 
test["Predicted_Profit"] = y_pred  
test
```

***MAE*** Ortalama mutlak hata iki sürekli değişken arasındaki farkın ölçüsüdür. MAE, her gerçek değer ile veriye en iyi uyan çizgi arasındaki ortalama dikey mesafedir.

```python
MAE = mean_absolute_error(y_test, y_pred)
MAE
```

***MSE*** bir regresyon eğrisinin bir dizi noktaya ne kadar yakın olduğunu söyler. Sıfıra yakın olması daha iyi performansı gösterir.

```python
MSE = mean_squared_error(y_test, y_pred)
MSE
```

***RMSE*** Regresyon hattının veri noktalarından ne kadar uzakta olduğunun bir ölçüsüdür.RMSE değerinin sıfır olması modelin hiç hata yapmadığı anlamına gelir

```python
RMSE = np.sqrt(MSE)
RMSE
```

***R-Squared*** Bağımsız değişkenlerin bağımlı değişkendeki değişikliği açıklama yüzdesidir. Bire ne kadar yaklaşırsa o kadar anlamlıdır
