**Analysing the impact of age, avg_glucose_level and bmi on stroke**

```python
sns.pairplot(data=data[['age','avg_glucose_level','bmi','stroke']],height=3, hue='stroke')
```

<font color='blue'>**OBSERVATION :** People with higher age and lower bmi are prone to stroke</font>

**Finding the heatmap of the correlation (corr) of the features in the 'data'**

```python
ax = plt.subplots(figsize=(12, 5));
ax = sns.heatmap(data.corr(), annot = True) ;
```

<font color='blue'>**OBSERVATION :** Age, hypertension, heart_disease, avg_glucose_level and ever_married has +ve correlation with the stroke </font>

**Analysing the impact of gender on stroke**

```python
pd.crosstab(data.stroke, data.gender, margins=True)
```

<font color='blue'>**OBSERVATION :** Females are more prone to stroke</font>

**Analysing the impact of ever_married status on stroke**

```python
pd.crosstab(data.stroke, data.ever_married, margins=True)
```

<font color='blue'>**OBSERVATION :** Married people are more prone to stroke</font>

**Analysing the impact of Residence_type on stroke**

```python
pd.crosstab(data.stroke, data.Residence_type, margins=True)
```

<font color='blue'>**OBSERVATION :** People living urban are more prone to stroke</font>

**Analysing the impact of hypertension on stroke**

```python
pd.crosstab(data.stroke, data.hypertension, margins=True)
```

<font color='blue'>**OBSERVATION :** People with NO hypertension are more prone to stroke</font>

**Analysing the impact of heart disease on stroke**

```python
pd.crosstab(data.stroke, data.heart_disease, margins=True)
```

<font color='blue'>**OBSERVATION :** People with NO heart disease are more prone to stroke</font>
