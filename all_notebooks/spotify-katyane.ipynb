## **Introdução**

Spotify é um serviço de streaming de música, podcast e vídeo que foi lançado oficialmente em 7 de outubro de 2008, onde fornece acesso a mais de 30 milhões de músicas. Nesse notebook foi analisado as 50 músicas mais ouvidas no ano de 2019.

**Dataset**

Os dados foram extraídos de:http://organizeyourmusic.playlistmachinery.com/.
O conjunto de dados dispõe de 50 musicas e 13 variáveis e 14 colunas sendo: 


Beats.Per.Minute - Batimentos por minuto, O andamento da música.

Energy - A energia de uma música - quanto maior o valor, mais energético. canção

Danceability - Quanto maior o valor, mais fácil é dançar essa música.

Loudness - Quanto maior o valor, mais alta a música

Liveness - Quanto maior o valor, maior a probabilidade de a música ser uma gravação ao vivo.

Valence - Quanto maior o valor, mais humor positivo para a música.

Length - A duração da música.

Acousticness - Quanto maior o valor, mais acústica é a música.

Speechiness - Quanto maior o valor, mais a palavra falada a música contém.

Popularity - Quanto maior o valor, mais popular é a música.

## **Perguntas**

1. Qual é a média da popularidade que um artista de topo tem?
2. Qual artista concentra a maior quantidade de música no top 50?
3. Qual o gênero concentra a maior quantidade de música no top 50?
4. Qual é a média de energia que um artista de topo tem?
5. Qual é a média de acústica que um artista de topo tem?
6. Qual é a média de humor positivo que um artista de topo tem?



## Codificação

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import plotly.graph_objects as go
```

```python
df = pd.read_csv('../input/top50spotify2019/top50.csv',encoding="ISO-8859-1")
```

```python
df.head()
```

```python
df.info()
```

![](http://)
Todas as variáveis são preenchidas, não existem objetos nulos.

```python
temp = df[['Artist.Name', 'Popularity']].groupby('Artist.Name', as_index=False).mean().sort_values(by='Popularity', ascending=False)[:10]
fig = go.Figure()
fig.add_trace(go.Bar(x=temp['Artist.Name'], y=temp['Popularity'], text=temp['Popularity'], textposition='auto'))
fig.update_layout(title='Qual é a média da popularidade que um artista de topo tem?')
fig.show()
```

Resposta da pergunta 1: Para poder observamos melhor a média de popularidade o gráfico foi limitado pelo top 10 e percebemos que Bad Bunny possui a maior média de popularidade neste conjunto de dados.


```python
temp = df[['Artist.Name', 'Genre']].groupby(['Artist.Name'], as_index=False).count().sort_values(by='Genre', ascending=False)[:3]
fig = go.Figure(data=[
    go.Bar(x=temp['Artist.Name'], y=temp['Genre'])
])
fig.update_layout(title='Qual artista concentra a maior quantidade de música no top 50?')
fig.show()
```

Resposta da pergunta 2: Para poder observamos melhor a média de popularidade o gráfico foi limitado pelo top 3 e percebemos que Ed Sheeran concentra a maior quantidade de música neste conjunto de dados.


```python
temp = df[['Genre', 'Liveness']].groupby(['Genre'], as_index=False).count().sort_values(by='Liveness', ascending=False)[:9]
fig = go.Figure()
fig.add_trace(go.Bar(x=temp['Genre'], y=temp['Liveness']))
fig.update_layout(title='Qual o gênero concetra a maior quantidade de musica no top 50?')
fig.show()
```

Resposta da pergunta 3: Dance Pop concentra a maior quantidade de gênero neste conjunto de dados.

```python
temp = df[['Artist.Name', 'Energy']].groupby('Artist.Name', as_index=False).mean().sort_values(by='Energy', ascending=False)[:10]
fig = go.Figure()
fig.add_trace(go.Bar(x=temp['Artist.Name'], y=temp['Energy'], text=temp['Energy'], textposition='auto'))
fig.update_layout(title='Qual é a média de energia que um artista de topo tem?')
fig.show()
```

![](http://)

Resposta da pergunta 4: Para poder observamos melhor a média de energia o gráfico foi limitado pelo top 10 e percebemos que Katy Perry possui a maior média de energia neste conjunto de dados.

```python
temp = df[['Artist.Name', 'Acousticness..']].groupby('Artist.Name', as_index=False).mean().sort_values(by='Acousticness..', ascending=False)[:10]
fig = go.Figure()
fig.add_trace(go.Bar(x=temp['Artist.Name'], y=temp['Acousticness..'], text=temp['Acousticness..'], textposition='auto'))
fig.update_layout(title='Qual é a média de acústica que um artista de topo tem?')
fig.show()
```

Resposta da pergunta 5: Para poder observamos melhor a média de acústica o gráfico foi limitado pelo top 10 e percebemos que Lewis Capaldi possui a maior média de acústica neste conjunto de dados.

```python
temp = df[['Artist.Name', 'Valence.']].groupby('Artist.Name', as_index=False).mean().sort_values(by='Valence.', ascending=False)[:10]
fig = go.Figure()
fig.add_trace(go.Bar(x=temp['Artist.Name'], y=temp['Valence.'], text=temp['Valence.'], textposition='auto'))
fig.update_layout(title='Qual é a média de humor positivo que um artista de topo tem?')
fig.show()
```

Resposta da pergunta 6: Para poder observamos melhor a média de energia o gráfico foi limitado pelo top 10 e percebemos que Jonas Brothers possui a maior média de humor positivo neste conjunto de dados.

```python
temp = df[['Artist.Name', 'Popularity', 'Energy', 'Acousticness..', 'Valence.']].groupby('Artist.Name', as_index=False).mean().sort_values(by='Popularity', ascending=False) [:20]

media = ['Popularity', 'Energy', 'Acousticness..', 'Valence.']

fig = go.Figure()

for me in media:
    fig.add_trace(go.Bar(x=temp['Artist.Name'], y=temp[me], name=me))
    
fig.update_layout(title='Conclusão do Top 20', barmode='stack', )
```

##Analise

Concluímos que observando os gráficos, Ed Sheeran é o artista que mais tem música no topo. E quando analisamos as questões separadamente fazendo a média das música, percebemos que  média de acústica do artista não aparece entre os top 20 do gráfico, mesmo obtendo a maior quantidade(Ed Sheeran), Bad Bunny lidera por ter uma maior média de acústica, então nem sempre o artista que está no topo ele vai ter a maior média de popularidade,a maior média de energia, a maior média de acústica e a maior média de humor positivo. Mais percebemos que a média do artistas de popularidade são similares oscilando apenas 1% entre eles.
