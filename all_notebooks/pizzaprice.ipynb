```python
import numpy as np
import scipy as sp
import pandas as pd

import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../input/predict-pizza-price/Pizza-Price.csv')
data
```

```python
data['Extra Cheeze'] = data['Extra Cheeze'].map({'yes':1, 'no':0})
data['Extra Mushroom'] = data['Extra Mushroom'].map({'yes':1, 'no':0})
data['Extra Spicy'] = data['Extra Spicy'].map({'yes':1, 'no':0})
```

```python
data.hist(figsize=(9,9), ec='k')
plt.tight_layout()
plt.show()
```

```python
data = pd.concat([pd.get_dummies(data['Restaurant']), data], axis=1).drop('Restaurant', axis=1)
```

```python
data
```

```python
data.info()
```

```python
data.describe()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data.drop('Price', axis=1), data['Price'], test_size=0.4, random_state=42)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error

clf = SGDRegressor()

clf.fit(x_train, y_train)
y_clf = clf.predict(x_test)
mean_squared_error(y_test, y_clf, squared=False)
```

```python
from sklearn.tree import DecisionTreeRegressor
clf_2 = DecisionTreeRegressor()

clf_2.fit(x_train, y_train)
y_clf_2 = clf_2.predict(x_test)
mean_squared_error(y_test, y_clf_2, squared=False)
```

```python
from sklearn.svm import SVR
clf_svr = SVR()

clf_svr.fit(x_train, y_train)
y_clf_svr = clf_svr.predict(x_test)
mean_squared_error(y_test, y_clf_svr, squared=False)
```
