```python
ACOESPERC_df = pd.read_csv('../input/correlao/ACOESFECH_2021.csv')
BESTCORRE_Cols_df = pd.read_csv('../input/correc2021-colunas/TODAS_CORR_COLUNAS_2021.csv')

display (ACOESPERC_df)
display (ACOESPERC_df.dtypes)
display (ACOESPERC_df.shape)

display (BESTCORRE_Cols_df)
display (BESTCORRE_Cols_df.dtypes)
display (BESTCORRE_Cols_df.shape)
```

```python
Temp = pd.DataFrame()
x = pd.DataFrame()
y = pd.DataFrame()

Temp['AZUL4'] = ACOESPERC_df['AZUL4']
x['AZUL4'] = ACOESPERC_df['AZUL4']

Temp['GOLL4'] = ACOESPERC_df['GOLL4']
y['GOLL4'] = ACOESPERC_df['GOLL4']

CORREL = Temp.corr()

display ('Correlação: ', CORREL)

R2 = CORREL **2

display ('R²: ', R2)

```

```python
from sklearn.linear_model import LinearRegression
Regres = LinearRegression()
Regres.fit(x, y)

print ('Coeficiente (1) Beta: ', Regres.coef_)
print ('Intercepto: ', Regres.intercept_)
print ('R²: ', Regres.score(x, y))
```

```python
def run_regression(df,function):
    model = smf.ols(function, df).fit()
    print(model.summary())

run_regression(ACOESPERC_df,'GOLL4 ~ AZUL4')
```

```python
run_regression(ACOESPERC_df,'AZUL4 ~ GOLL4 + ALSO3 + SUZB3 + CCRO3 + ECOR3')
```
