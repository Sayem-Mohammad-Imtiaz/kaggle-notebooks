```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/data-police-shootings/fatal-police-shootings-data.csv')
df
```

```python
df.dtypes
```

```python
year=[]
for i in range(len(df)):
    year.append(df['date'][i].split('-')[0])
year
```

```python
df['year']=year
df
```

```python
df1=df.groupby('year')['id'].count()
df1
```

```python
df1=df1.reset_index()
df1
```

# Plot to visualise rate of shootings

```python
plt.plot(df1['year'],df1['id'],'g+-')
plt.ylim(ymin=0)
```

```python
df2=df.groupby('race')['id'].count()
df2
```

```python
df2=df2.reset_index()
df2
```

# Rate of killings relative to race


```python
plt.bar(df2['race'],df2['id'])
```


W: White, non-Hispanic B: Black, non-Hispanic A: Asian N: Native American H: Hispanic O: Other
