```python
plt.figure(figsize=(14,8))
sns.barplot(data=df,x='State',y='Number of Test Centers',color=sns.color_palette('Set3')[0])
plt.title('Number of Test Centers')
plt.xlabel('State')
plt.ylabel('Number of Test Centers')
plt.xticks(rotation=45)
for i in range(df.shape[0]):
    count = df.iloc[i]['Number of Test Centers']
    plt.text(i,count+1,df.iloc[i]['Number of Test Centers'],ha='center')
    
from IPython.display import display, Markdown
display(Markdown("Most Number of cases have occured in the Test Centers **20-50**"))
```

```python
df_grp = df.groupby(["Days from April 15th","Peak Date"])[["State","Median Income","Population", "Number of Test Centers", "Total Cases"]].sum().reset_index()
df_grp.head()
```

```python
df_grp = df_grp.rename(columns={"Days from April 15th":"days","Peak Date":"peak", "Median Income": "income", "Number of Test Centers": "test", "Total Cases": "total"})
```

```python
plt.figure(figsize=(15, 5))
plt.title('test')
df_grp.test.value_counts().plot.bar();
```

```python
df_grp_plot = df_grp.tail(80)
```

```python
fig=px.bar(df_grp_plot,x='peak', y="total", animation_frame="days", 
           animation_group="peak", color="peak", hover_name="peak")
fig.update_yaxes(range=[0, 1500])
fig.update_layout(title='Peak Date')
```

```python
fig = px.bar(df_grp[['test', 'total']].sort_values('total', ascending=False), 
             y="total", x="test", color='test', 
             log_y=True, template='ggplot2', title='Total Cases vs Tests')
fig.show()
```

```python
df_grp = df_grp.sort_values(by=['test'],ascending = False)
```

```python
plt.figure(figsize=(40,15))
plt.bar(df_grp.test, df_grp.total,label="total")
plt.bar(df_grp.test, df_grp.peak,label="peak")
plt.bar(df_grp.test, df_grp.days,label="days")
plt.xlabel('Peak Date')
plt.ylabel("Total Cases")
plt.xticks(fontsize=13)
plt.yticks(fontsize=15)

plt.legend(frameon=True, fontsize=12)
plt.title('Total Cases vs Peak Date vs Test Centers',fontsize=30)
plt.show()

f, ax = plt.subplots(figsize=(40,15))
ax=sns.scatterplot(x="test", y="total", data=df_grp,
             color="black",label = "Confirm")
ax=sns.scatterplot(x="test", y="peak", data=df_grp,
             color="red",label = "peak")
ax=sns.scatterplot(x="test", y="days", data=df_grp,
             color="blue",label = "days")
plt.plot(df_grp.test,df_grp.total,zorder=1,color="black")
plt.plot(df_grp.test,df_grp.peak,zorder=1,color="red")
plt.plot(df_grp.test,df_grp.days,zorder=1,color="blue")
plt.xticks(fontsize=13)
plt.yticks(fontsize=15)
plt.legend(frameon=True, fontsize=12)
```

```python
fig = px.line(df_grp, x="peak", y="total", 
              title="Covid-19 Peak date & Total cases ")
fig.show()
```
