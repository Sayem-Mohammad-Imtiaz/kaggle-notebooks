# Exploritary data analysis

# Books


Top 10 mentioned books

```python
books_count = df['Name'].value_counts().reset_index().head(10).rename(columns={'index':'Name', 'Name':'Count'})

fig, ax = plt.subplots()
ax.barh(books_count['Name'], books_count['Count'])
ax.set_yticks(books_count['Name'])
ax.set_yticklabels(books_count['Name'])
ax.invert_yaxis()
plt.grid(True)
ax.set_xlabel('Count')
ax.set_ylabel('Name')
ax.set_title('Books count')
plt.show()
```

```python
author_count = df['Author'].value_counts().reset_index().head(10).rename(columns={'index':'Author', 'Author':'Count'})

fig, ax = plt.subplots()
ax.barh(author_count['Author'], author_count['Count'])
ax.set_yticks(author_count['Author'])
ax.set_yticklabels(author_count['Author'])
ax.invert_yaxis()
plt.grid(True)
ax.set_xlabel('Count')
ax.set_ylabel('Author')
ax.set_title('Authors count')
plt.show()
```

# User rating

User rating distribution

```python
fig = plt.figure(figsize = (10, 10))
sns.distplot(df['User Rating'])
plt.axvline(df['User Rating'].mean(), color = 'red')
plt.grid(True)
plt.title('Distribution of User rating')
plt.show()
```

User rating distrbution by genre

```python
fig = plt.figure(figsize = (10, 10))

sns.distplot(df.loc[df['Genre']=='Fiction']['User Rating'], label = 'Fiction')
plt.axvline(df.loc[df['Genre']== 'Fiction']['User Rating'].mean(), color = 'blue')

sns.distplot(df.loc[df['Genre']=='Non Fiction']['User Rating'], label = 'Non Fiction')
plt.axvline(df.loc[df['Genre']== 'Non Fiction']['User Rating'].mean(), color = 'orange')
plt.grid(True)
plt.title('Distribution of user rating by genre')
plt.legend()
plt.show()
```

We can see genre fiction has higher user rating than non fiction

# Reviews

Reviews distribution

```python
fig = plt.figure(figsize = (10, 10))
sns.distplot(df['Reviews'])
plt.axvline(df['Reviews'].mean())
plt.grid(True)
plt.title('Reviews ditribution')
plt.show()
```

Reviews distribuition by genre

```python
fig = plt.figure(figsize = (10, 10))

sns.distplot(df.loc[df['Genre']=='Fiction']['Reviews'], label = 'Fiction')
plt.axvline(df.loc[df['Genre']== 'Fiction']['Reviews'].mean(), color = 'blue')

sns.distplot(df.loc[df['Genre']=='Non Fiction']['Reviews'], label = 'Non Fiction')
plt.axvline(df.loc[df['Genre']== 'Non Fiction']['Reviews'].mean(), color = 'orange')
plt.grid(True)
plt.title('Distribution of price by genre')
plt.legend()
plt.show()
```

We can see that fiction reviews are more distributed

# Price

Price distribution

```python
fig = plt.figure(figsize = (10, 10))
sns.distplot(df['Price'])
plt.axvline(df['Price'].mean())
plt.grid(True)
plt.title('Distribution of price')
plt.show()
```

Price distribution by genre

```python
fig = plt.figure(figsize = (10, 10))

sns.distplot(df.loc[df['Genre']=='Fiction']['Price'], label = 'Fiction')
plt.axvline(df.loc[df['Genre']== 'Fiction']['Price'].mean(), color = 'blue')

sns.distplot(df.loc[df['Genre']=='Non Fiction']['Price'], label = 'Non Fiction')
plt.axvline(df.loc[df['Genre']== 'Non Fiction']['Price'].mean(), color = 'orange')
plt.grid(True)
plt.title('Distribution of price by genre')
plt.legend()
plt.show()
```

# Year

How many books were printed during each year by genre

```python
fig = plt.figure(figsize = (10, 10))
sns.countplot(x = df['Year'], hue = 'Genre', data = df)
plt.xlabel('Year')
plt.ylabel('Count')
plt.grid(True)
plt.title('Books released each year by genre')
plt.show()
```

# Genre

```python
plt.figure(figsize = (20,10))
sns.countplot(y = 'Genre', data = df)
plt.grid(True)
plt.title('Genre count')
plt.show()
```

# Most popular books by user rating

```python
top10_books_user_rating = df.groupby('Name')['User Rating'].mean().reset_index().head(10).sort_values(by = 'User Rating', ascending = False)

fig, ax = plt.subplots()
ax.barh(top10_books_user_rating['Name'], top10_books_user_rating['User Rating'])
ax.set_yticks(top10_books_user_rating['Name'])
ax.set_yticklabels(top10_books_user_rating['Name'])
ax.invert_yaxis()
plt.grid(True)
ax.set_xlabel('User Rating')
ax.set_ylabel('Book')
ax.set_title('User ratings of different books')
plt.show()
```

# Which books has highest reviews

```python
top10_book_reviews = df.groupby('Name')['Reviews'].mean().reset_index().head(10).sort_values(by = 'Reviews', ascending = False)

fig, ax = plt.subplots()
ax.barh(top10_book_reviews['Name'], top10_book_reviews['Reviews'])
ax.set_yticks(top10_book_reviews['Name'])
ax.set_yticklabels(top10_book_reviews['Name'])
ax.invert_yaxis()
ax.set_xlabel('Reviews')
plt.grid(True)
ax.set_ylabel('Book')
ax.set_title('Reviews of different books')
plt.show()
```

# Most expensive books

```python
top_expensive_books = df.groupby('Name')['Price'].mean().reset_index().head(10).sort_values(by = 'Price', ascending = False)

fig, ax = plt.subplots()
ax.barh(top_expensive_books['Name'], top_expensive_books['Price'])
ax.set_yticks(top_expensive_books['Name'])
ax.set_yticklabels(top_expensive_books['Name'])
ax.invert_yaxis()
plt.grid(True)
ax.set_xlabel('Price')
ax.set_ylabel('Book')
ax.set_title('Most expensive books')
plt.show()
```
