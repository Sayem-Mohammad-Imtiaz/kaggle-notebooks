```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly
```

```python
df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df.index=pd.to_datetime(df['date'])
df=df.sort_index()
print(df.columns)
fig=px.choropleth(df,locations='country',locationmode='country names',color='total_vaccinations_per_hundred',hover_name='country',animation_frame='date',color_continuous_scale='Oranges',range_color=(0.0,80,0))
fig.update_layout(title_text="percentage of country's total population that are fully vaccinated")
fig.show()

plotly.offline.plot(fig,filename='vaccinatons.html')
```

This figure shows that the situation of vaccination all of the world. I tried to show the degree of the situation of vaccination by the intensity of the color.
