# Step 1: Primary checks and data cleanup.

```python
print("\n Shape of dataset : ",players_health.shape)
```

```python
#listing down uniques, for now I want to drop the 'date' column and see how EDA and modelling goes.

print("\n Unique notes : ",len(players_health['Notes'].unique()))
print("\n Unique names : ",len(players_health['Name'].unique()))
print("\n Unique teams : ",len(players_health['Team'].unique()))
print("\n Unique positions : ",len(players_health['Position'].unique()))

players_health = players_health.drop(columns=['Date'])
```

```python
# label encoder for all string columns.

from sklearn.preprocessing import LabelEncoder
current_encoder = LabelEncoder()


# Encoding just the notes column into a Injury based column.
players_health['Injury_Type'] = current_encoder.fit_transform(players_health['Notes'])

players_health = players_health.drop(columns=['Notes'])

# Dropping 'Notes' now.

players_health.head(10)
```

# Step 2: Visualizing the numeric dataframe.

```python
# Starting with a few barplots.

team_vs_injury = players_health.groupby('Team')['Injury_Type'].sum()
colors = range(41,71)
barplot1 = plex.bar(x=team_vs_injury.index,y=team_vs_injury.values, title="Teams with injured players(sum totals)",color=colors)
barplot1.show()
```

```python
position_vs_injury = players_health.groupby('Position')['Injury_Type'].sum()
colors = range(18,11,-1)
barplot1 = plex.bar(x=position_vs_injury.index,y=position_vs_injury.values, title="Positions with injured players(sum totals)",color=colors)
barplot1.show()
```

```python
# Before setting up a correlation heatmap. Encoding remaining columns.

players_health['Position'] = current_encoder.fit_transform(players_health['Position'])
players_health['Team'] = current_encoder.fit_transform(players_health['Team'])
players_health['Name'] = current_encoder.fit_transform(players_health['Name'])

players_health.head(10)
```
