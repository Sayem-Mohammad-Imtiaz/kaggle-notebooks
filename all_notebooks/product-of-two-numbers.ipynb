```python
np.random.seed(1)
```

```python
def create_train_data(low, high, train_examples):

    x_train = []
    y_train = []

    a = np.random.randint(low=low, high=high, size=train_examples, dtype='int')
    b = np.random.randint(low=low, high=high, size=train_examples, dtype='int')

    c = np.multiply(a,b)

    x_train = pd.concat([pd.Series(a),pd.Series(b)], axis=1)
    y_train = pd.DataFrame(c)

    return x_train, y_train
```

```python
def create_test_data(low, high, test_examples):
    
    x_test=[]
    y_test=[]
    
    a = np.random.randint(low=low, high=high, size=test_examples, dtype='int')
    b = np.random.randint(low=low, high=high, size=test_examples, dtype='int')

    x_test = pd.concat([pd.Series(a),pd.Series(b)], axis=1)
    c = np.multiply(a,b)

    y_test = pd.DataFrame(c)

    return x_test, y_test 
```

```python
x_train, y_train = create_train_data(10, 99, 50000)
x_test, y_test = create_test_data(10, 99, 10000)
```

```python
import tensorflow as tf
import matplotlib.pyplot as plt
```

## Model1

```python
tf.random.set_seed(15)
```

```python
model1 = tf.keras.models.Sequential()
model1.add(tf.keras.layers.Dense(1, input_dim=2, activation='relu'))
model1.summary()
```

```python
model1.compile(loss='mean_squared_error', optimizer='adam')
callback = tf.keras.callbacks.EarlyStopping(monitor="loss", min_delta=1e-4,patience=20)
```

```python
history = model1.fit(x_train, y_train, epochs=250, validation_data=(x_test, y_test), callbacks=[callback])
train_mse = model1.evaluate(x_train, y_train, verbose=0)
test_mse = model1.evaluate(x_test, y_test, verbose=0)
plt.title('Loss / Mean Squared Error')
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```

```python
model1.evaluate(x_test, y_test)
```

```python
model1.metrics_names
```

```python
pred_train= model1.predict(x_train)
rmse = tf.keras.metrics.RootMeanSquaredError()
rmse.update_state(y_train,pred_train)
print("Train RMSE Score for Model1: ")
rmse.result().numpy()
```

```python
pred_test= model1.predict(x_test)
rmse.reset_states()
rmse.update_state(y_test,pred_test)
print("Test RMSE Score for Model1: ")
rmse.result().numpy()
```
