```python
from pathlib import Path
#from sklearn.preprocessing import LabelEncoder
import pandas as pd

PATH_TRAIN = Path("../input/stanford-natural-language-inference-corpus/snli_1.0_train.csv")
PATH_TEST = Path("../input/stanford-natural-language-inference-corpus/snli_1.0_test.csv")

## reading data
df_train = pd.read_csv(PATH_TRAIN)
df_test = pd.read_csv(PATH_TEST)
```

```python
#training data preprocessing
del df_train['sentence1_binary_parse']
del df_train['sentence2_binary_parse']
del df_train['sentence1_parse']
del df_train['sentence2_parse']
del df_train['captionID']
del df_train['pairID']
del df_train['label1']
del df_train['label2']
del df_train['label3']
del df_train['label4']
del df_train['label5']

df1 = df_train.reindex(columns= ['sentence1', 'sentence2', 'gold_label'])
df1.columns = ['premise','hypothesis','label']
df1['label'].value_counts()
df1 = df1[df1.label != '-']

#remove missing values
df1 = df1.dropna()

#encode labels
numeric_labels = {"entailment": 0, "neutral": 1, "contradiction": 2}
df1['label'] = df1.label.map(numeric_labels)

#export as csv
df1.to_csv("train.csv", index=False)
```

```python
#testing data preprocessing
del df_test['sentence1_binary_parse']
del df_test['sentence2_binary_parse']
del df_test['sentence1_parse']
del df_test['sentence2_parse']
del df_test['captionID']
del df_test['pairID']
del df_test['label1']
del df_test['label2']
del df_test['label3']
del df_test['label4']
del df_test['label5']

df2=df_test.reindex(columns= ['sentence1', 'sentence2', 'gold_label'])
df2.columns = ['premise','hypothesis', 'label']
df2['label'].value_counts()
df2 = df2[df2.label != '-']

#remove missing values
df2 = df2.dropna()

del df2['label']
df2.to_csv("test.csv", index=False)
```
