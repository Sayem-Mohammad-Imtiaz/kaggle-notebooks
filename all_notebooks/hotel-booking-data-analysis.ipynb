**Introduction**
Have you ever wondered the trends for hotel bookings? How long people stay On average? Most travelers from which countries?How many children did the bookers have? In this analysis I explore a large dataset to examine these questions.

**Dataset context**
hotel_booking  dataset contains 119390 observations for a City Hotel and a Resort Hotel. Each observation represents a hotel booking between the 1st of July 2015 and 31st of August 2017, including booking that effectively arrived and booking that were canceled.

**Acknowledgements**
The data is originally from the article Hotel Booking Demand Datasets, written by Nuno Antonio, Ana Almeida, and Luis Nunes for Data in Brief, Volume 22, February 2019.

```python
#Q7 =name of the aperson who has  max ADR 

max_adr = df ['adr'].idxmax()
name = df.iloc[max_adr]['name']
max_adr=df.iloc[max_adr]['adr']
print('the max of adr is ({:.2f}) for (({}))' .format(max_adr,name))

```

```python
#Q8 =  average of all payment

mean_adr = np.round(df['adr'].mean(),2)
print('mean ADR is {}.'.format(mean_adr))
```

```python
#Q9 = mean nights

df['total_night'] = df['stays_in_weekend_nights']+df['stays_in_week_nights']
Mean_total_night = df['total_night'].mean()
print('Mean nights : ', round(Mean_total_night,2))
```
