```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

## Hypothesis #1:
Students with previous attempts in a course are likely to pass than a student taking the course first time.

**Resoning**: My assumption is that students who took a course previously have some background knowledge of that course. Hence, they are likely to pass the course.**

```python
studentInfo = pd.read_csv('/kaggle/input/student-demographics-online-education-dataoulad/studentInfo.csv')
# students with previous attempts
studentsWithPreviousAttempts = studentInfo[studentInfo.num_of_prev_attempts > 0]
studentsWithPreviousAttempts = studentsWithPreviousAttempts['final_result'].value_counts()
studentsWithPreviousAttempts.plot(kind='bar')
```

```python
# students without previous attempts
studentsWithOutPreviousAttempts = studentInfo[studentInfo.num_of_prev_attempts == 0]
studentsWithOutPreviousAttempts = studentsWithOutPreviousAttempts['final_result'].value_counts()
studentsWithOutPreviousAttempts.plot(kind='bar')
```

To calculate students with previous attempts, I'll use table `StudentInfo` and get students with `num_of_prev_attempts` equal to zero and greater than 0 separately. Than based on the data, we can calculate the count of `final_result` values and plot them.  

If we see the two different plots, we can see that my hypothesis is wrong. Students who didn't take courses previously actually passed more frequently. 

**Alternative interpretation**: Students with previous attempts passing less frequently may indicate that they don't focus on the course materials again as well and also may be they are not interested in those courses at all.  

## Hypothesis #2:
Students studying more credits at a time are more likely to withdraw courses than students taking less credits.

**Resoning**: My assumption is that students who are studying more credits have more workload to manage that's why they withdraw frequently.

```python
studentsWithMoreCredits = studentInfo[studentInfo.studied_credits > 150]
studentsWithMoreCredits = studentsWithMoreCredits['final_result'].value_counts()
studentsWithMoreCredits.plot(kind='bar')
```

```python
studentsWithLessCredits = studentInfo[studentInfo.studied_credits <= 150]
studentsWithLessCredits = studentsWithLessCredits['final_result'].value_counts()
studentsWithLessCredits.plot(kind='bar')
```

To calculate students studying more credits, I'll use table `StudentInfo` and get students with `studied_credits` less than and equal to 150 and greater than 150 separately. Than based on the data, we can calculate the count of `final_result` values and plot them.  

If we look at the two different plots, we can see that my hypothesis is correct. Students who are studying more courses are withdrawing more frequently. 

**Alternative interpretation**: May be students are withdrawing because they don't like the course.  

## Hypothesis #3:
Students registering late to the courses are unlikely to passthan students who are registering early.

**Resoning**: My assumption is that students who are registering late will be left behind and they have to catch up. Which is not easy always and they may miss some assignments as well.

```python
studentRegistration = pd.read_csv('/kaggle/input/student-demographics-online-education-dataoulad/studentRegistration.csv')
studentRegistrationWithInfo = studentRegistration.join(studentInfo.set_index('id_student'), on='id_student', how='left', lsuffix='_left', rsuffix='_right')

studentsWithLateRegistration = studentRegistrationWithInfo[studentRegistrationWithInfo.date_registration > 30]
studentsWithLateRegistration['final_result'].value_counts().plot(kind='bar')
```

```python
studentsWithEarlyRegistration = studentRegistrationWithInfo[studentRegistrationWithInfo.date_registration <= 30]
studentsWithEarlyRegistration['final_result'].value_counts().plot(kind='bar')
```

To calculate students registering late, I'll use table `studentRegistration` and join it with `studentInfo` to get result at one place. After than, we can extract data with `date_registration` greater than 30 and less than and equal to 30 separately. Than based on the data, we can calculate the count of `final_result` values and plot them.  

If we see the two different plots, we can see that my hypothesis is kind of correct. Students who registered late actually failed or withdrew more frequently. 

**Alternative interpretation**: Students registering late may already have knowledge of that course.
