```python
##convert year to datetime

video["Year_of_Release"] = pd.to_datetime(video["Year_of_Release"], format='%Y')


##sort by date for timeseries

video = video.sort_values(by='Year_of_Release') 

video["year"] = video["Year_of_Release"].dt.year

```

## Video Analysis


> ### Sales per Genre

Which genres account for the majority of the sales? Does it differ between countries?

```python
sales = ["NA_Sales","EU_Sales","JP_Sales","Global_Sales","Other_Sales"]
```

```python
## make a for loop for efficiency

for i in sales:

    plt.figure(figsize=(15,16))
    ax = sns.barplot(x=i, y="Genre", hue="Genre",
                     data=video,dodge=False)
    plt.title(f"{i} per Genre")
    plt.xlabel('Sales (M)')
    plt.ylabel('Genre')
```

- North America and Europe receive the majority of their sales from the shooting and platform genres, whereas Japan's majority of sales comes from the role-playing genre
- With the release of many popular platform games like Donkey Kong, and Mario bros , its not suprising that platform contributes to a large amount of sales

- In regards to Japan the release of the Final Fantasy series and Pokemon played a large part in the high interest in Role-Playing genre. There must also be an influence of Japanese Anime culture on the choice of Role-Playing games as the majority of Role-Playing are Anime related games such as Pokemon, Yu-Gi-Oh and Dragonball-Z.


- Global sales show that platform is the highest grossing genre overall followed by shooter 

How does the popularity of genres change over time?

```python

plt.figure(figsize=(15,7))
ax = sns.barplot(x="year", y="Global_Sales", hue="Genre",
                 data=video, dodge=False)
ax.legend(loc="upper right")
plt.title("Global Sales per Genre")
plt.xlabel('Release Year')
plt.ylabel('Sales (M)')
plt.xticks(rotation='vertical')
plt.yticks(rotation='vertical');
```

- The Platform genre dominates the gaming industry during the 80's with games such as Mario Bros, 
- ~1986 the role-playing genre becomes the most popular and continues to contribute to the majority of sales until 2016

### Sales Per Country 

```python
## groupby year and aggregate by sum

year_sum = video.groupby(["year"]).agg({"sum"}).reset_index()
```

```python
fig, ax = plt.subplots(figsize=(16,9))
ax.plot(year_sum["year"],
        year_sum["NA_Sales"],
        color="r",
        label="North American Sales");
ax.plot(year_sum["year"],
        year_sum["EU_Sales"],
        color="b",
        label="Europe Sales");
ax.plot(year_sum["year"],
        year_sum["JP_Sales"],
        color="m",
        label="Japan Sales");
ax.plot(year_sum["year"],
        year_sum["Other_Sales"],
        color="k",
        label="Other Sales");

ax.set_title("Sales per Release Year")
plt.xlabel('Release_year')
plt.ylabel('Sales (M)')
ax.legend();
```

- ~2010 was where social gaming, cloud gaming, free to play and mobile apps start to reshape the gaming industry
- as you can see from the graph 2010 is where the sales start to decline steadily for all the countries.

Lets investigate why Japan has relatively less sales then Northern America and Europe ?

```python
def first_j(video):
    return video.sort_values(by='JP_Sales')[-3:]
```

```python
top_j = video.groupby(['year'], group_keys=False).apply(first_j)[['year', 'Name', 'JP_Sales', 'Genre', 'Platform', 'Publisher']]
```

### Sales per Publisher


What are the top grossing games per year?

```python
## define function to apply for aggregate application

def first(video):
    return video.sort_values(by='Global_Sales')[-1:]
```

```python
top = video.groupby(['year'], group_keys=False).apply(first)[['year', 'Name', 'Global_Sales', 'Genre', 'Platform', 'Publisher']]
```

```python
top
```

- The highest grossing publisher overall comes as no suprise, Nintendo taking out the top 5 spots globally with Wii Sports, Super Mario Bros, Duck Hunt, Wii Sports Resort, and Mario Kart Wii

Lets take a deeper look into Nintendo

```python
## subset for Nintendo 

n = top[top.Publisher == "Nintendo"]
```

```python
### calculate percentage of overall sales

s = n["Global_Sales"].sum()
t = video["Global_Sales"].sum()
u = s/t
```
