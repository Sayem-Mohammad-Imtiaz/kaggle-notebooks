```python
data = pd.read_csv("/kaggle/input/breast-cancer-wisconsin-data/data.csv")
data
```

```python
data.drop(["id","Unnamed: 32"],axis=1,inplace=True)
data
```

```python
"""
for t,each in enumerate(data.diagnosis):
    if each == "M":
        data.diagnosis[t] = 1
    else:
        data.diagnosis[t] = 0
"""

data.diagnosis = [1 if each == "M" else 0 for each in data.diagnosis]
data
```

```python
import matplotlib.pyplot as plt
hasta = data[data.diagnosis == 1]
normal = data[data.diagnosis == 0]
normal
```

```python
plt.figure(figsize=(12,12))
plt.scatter(hasta.radius_mean,hasta.compactness_mean,color="red",label="hasta",alpha=0.4)
plt.scatter(normal.radius_mean,normal.compactness_mean,color="blue",label="mormal",alpha=0.4)
plt.xlabel("Radius Mean")
plt.ylabel("Compactness Mean")
plt.legend()
plt.show()
```

```python
y = data.diagnosis.values
x = data.drop(["diagnosis"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=256,activation="sigmoid",input_dim=x.shape[1]))
model.add(Dense(units=512,activation="sigmoid"))
model.add(Dense(units=1,activation="sigmoid"))

model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])

model.fit(x_train,y_train,epochs=35)
```

```python
predicted = model.predict(x_test)
predicted
```

```python
for t,e in enumerate(predicted):
    if(predicted[t]<0.5):
        predicted[t] = int(0)
    else:
        predicted[t] = int(1)

predicted
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(predicted,y_test))
```
