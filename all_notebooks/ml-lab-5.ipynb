# Train|Test Split and Scaling Data

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
x = df.drop('Cancer Present',axis=1)
y = df['Cancer Present']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2,random_state=42)
```

**Data is split into 80% for training and 20% testing**

```python
scaler = StandardScaler()
```

```python
scaled_x_train = scaler.fit_transform(x_train)
scaled_x_test = scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=55)
knn_model.fit(scaled_x_train,y_train)
```

# Understanding KNN and Choosing K Value

```python
full_test = pd.concat([x_test,y_test],axis=1)
print(len(full_test))
sns.scatterplot(x='Gene One',y='Gene Two',hue='Cancer Present',data=full_test,alpha=0.7,palette='magma')
```

# Model Evaluation

```python
y_pred = knn_model.predict(scaled_x_test)
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
