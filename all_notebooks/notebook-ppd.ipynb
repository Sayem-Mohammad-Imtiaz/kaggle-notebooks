```python
# open auto complete
%config Completer.use_jedi = True

# data analysis and wrangling
import pandas as pd
import numpy as np
import random as rnd

# # visualization
# import seaborn as sns
# import matplotlib.pyplot as plt
# %matplotlib inline


# machine learning
from sklearn.svm import SVC #4
from sklearn.neighbors import KNeighborsClassifier #7
from sklearn.linear_model import SGDClassifier #9
from sklearn.tree import DecisionTreeClassifier #5

from sklearn.neural_network import MLPClassifier  #3
from sklearn.model_selection import train_test_split  # split train test size
from sklearn.naive_bayes import BernoulliNB, GaussianNB  #1
from sklearn.metrics import confusion_matrix  # metrics
from sklearn.ensemble import RandomForestClassifier, VotingClassifier  #2
from sklearn.ensemble import GradientBoostingClassifier #6
from imblearn.over_sampling import SMOTE, ADASYN
```

```python
train_0=pd.read_csv('../input/ppd-for-xuecheng/PPDtrain-en.csv')
test_0=pd.read_csv('../input/ppd-for-xuecheng/PPDtest-en.csv')
test_0
```

```python
drops=[]
for i in train_0:
    if train_0[i].dtype not in [np.float,np.int]:
        drops.append(i)
print(drops)
train = train_0.drop(drops,1)
test = test_0.drop(drops,1)

#train_0
train
```

```python
for i in train.keys():
    train[i].fillna(train[i].median(),inplace=True)
    test[i].fillna(test[i].median(),inplace=True)
#print(train.isna())
#print(train.isna().mean())
print(train.isna().mean().mean())
```

```python
id_train=train.pop('Idx')
id_test=test.pop('Idx')

train_y0 =train.pop('target')
test_y =test.pop('target')

id_train
#id_test
#train_y0
#train
```

```python
np.random.seed(100)
train_x,valid_x,train_y,valid_y = train_test_split(train,train_y0,test_size=0.3)
#print(train_x)
valid_x
```

```python
#resample
m_resample = SMOTE()
print(train_y.__len__())
train_x2,train_y2 = m_resample.fit_resample(train_x,train_y)
print(train_y2.__len__())
#train_x
train_x2
```

```python
#
valid_bench = np.zeros(valid_y.shape)
valid_conf = confusion_matrix(valid_y,valid_bench)
print(valid_conf)
test_bench = np.zeros(test_y.shape)
test_conf = confusion_matrix(test_y,test_bench)
print(test_conf)
```

```python
def roi(conf,ratio=[0.1,-1]):
    return np.mean(conf[:,0]*ratio)
```

```python
print(roi(valid_conf))
print(roi(test_conf))
```

```python
#1
m = BernoulliNB()
m.fit(train_x,train_y)
valid_pred1 = m.predict(valid_x)
valid_p_conf1 = confusion_matrix(valid_y,valid_pred1)
print(valid_p_conf1)
print(roi(valid_p_conf1))
```
