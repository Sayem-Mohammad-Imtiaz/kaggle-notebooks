**Price prediction of IT Secotr, Financials Sector and Health Care Sector**

**1.Information Technology Sector**

```python
sns.heatmap(df_IT.corr())
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
x = df_IT.drop(['Price','Symbol','Name','Sector','SEC Filings'],axis=1)
y = df_IT['Price']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=123)
```

```python
lr.fit(x_train,y_train)
```

```python
print(lr.intercept_)
print(lr.coef_)
```

```python
prediction_IT = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
print('MAE:',mean_absolute_error(y_test, prediction_IT))
print('MSE:',mean_squared_error(y_test, prediction_IT))
print('RMSE:',np.sqrt(mean_squared_error(y_test, prediction_IT)))
print('R2:',r2_score(y_test, prediction_IT))
```
