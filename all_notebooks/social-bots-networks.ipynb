#It requires more than one line to avoid zero words errors (not short sentences).

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
nlp = spacy.load("en_core_web_lg")  # make sure to use larger model!
tokens = nlp("I'll probably be hated for this but here goes. Multiple changes denotes insecurity. In my opinion all scream Hey look at me. I imagine attention is a good thing. But only for the right reason. What validate you? Some are in excess of a thousand dollars, to request what is amazing. I had the same thoughts about her. Nothing more attractive than ones NATURAL BEAUTY. Someone will definitely be fired over this, and very well should be.")

newText =''
for word in tokens:
 if word.pos_ in ['ADJ', 'NOUN']:
  newText = " ".join((newText, word.text.lower()))

wordcloud = WordCloud(stopwords=STOPWORDS, colormap='Set3', background_color="black").generate(newText)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
