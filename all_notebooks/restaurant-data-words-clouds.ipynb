```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS
stopwords = set(STOPWORDS)
```

```python
Data=pd.read_csv('../input/restaurant-data/RestaurantData.csv')
```

```python

print(Data.shape,len(Data['Unnamed: 0']))
Data.head()
```

```python
Data=Data.drop(columns=['Unnamed: 0'])
Data.head()
```

```python
def get_element(mylist,position):
    return mylist[position]
Data['Rating']=Data.Rating.str.split('of').apply(get_element,position=0)
Data.head()
```

```python
Data[['First','Restaurant Name']] = Data['Restaurant Name'].str.split(".",expand=True,)
Data=Data.drop(columns=['First'])
```

```python
Data.head()
```

```python
Data['Location'].value_counts().sort_index()
```

```python
 Agra=Data[Data.Location == 'Agra']
Bengaluru=Data[Data.Location == 'Bengaluru']
 Chennai=Data[Data.Location == 'Chennai']
Goa=Data[Data.Location == 'Goa']
 Hyderabad=Data[Data.Location == 'Hyderabad']
Jaipur=Data[Data.Location == 'Jaipur']
 Mumbai=Data[Data.Location == 'Mumbai']
Noida=Data[Data.Location == 'Noida']
 Patna=Data[Data.Location == 'Patna']
Pune=Data[Data.Location == 'Pune']
Shimla=Data[Data.Location == 'Shimla']
Srinagar=Data[Data.Location == 'Srinagar']
Udaipur=Data[Data.Location == 'Udaipur']
Varanasi=Data[Data.Location == 'Varanasi']
```

```python
wordcloud = WordCloud(
                          background_color='#42f5e3',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=40, 
                          random_state=42
                         ).generate(str(Agra['Cuisine']))

print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.title('Words Clouds For City Agra')
plt.show()
```
