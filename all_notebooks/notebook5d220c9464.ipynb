Let's clean our data now . Check null Values

```python
df.isna().sum()
```

There are none null values in our selected Dataset

Check how many values contain zeros in columns

```python
featureList = ['Glucose', 'BloodPressure',	'SkinThickness',	'Insulin',	'BMI']
print(df[featureList].isin({0}).sum())
```

Let's replace zeros with mean values

```python
for col in featureList:
  df[col] = df[col].replace({ 0 : df[col].mean() })
```

```python
featureList = ['Glucose', 'BloodPressure',	'SkinThickness',	'Insulin',	'BMI']
print(df[featureList].isin({0}).sum())
```

Now we have successfully replaced all zeros with suitable mean values . Now we have a cleaned dataset

Now Let's import some data visualization packages

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
plt.figure(figsize=(10, 5))
sns.barplot('Outcome', 'Pregnancies', data=df, ) 
plt.title("Bars of Outcome and Pregnancies")
plt.xlabel("Outcome")
plt.ylabel("Pregnancies")
plt.show()
```

```python
feature_columns = df[['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]

feature_columns.head()
```

```python
outcome_column = df['Outcome']
outcome_column.head()
```

Now let us Train our Data for Model Starting with importing packages

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split( feature_columns, outcome_column, test_size=0.2, random_state=5) 
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Now Let's fit our Model

```python
model = LogisticRegression()
```

```python
model = model.fit(X_train, y_train)

score = model.predict(X_train)
```

Let's Test Our Model Now

```python
print("Training Score: ", model.score(X_train, y_train))
print("Testing Score:  ", model.score(X_test, y_test))
```

```python
pred = model.predict(X_test)
print("Model Accuracy is : ", pred)
```

```python
model.coef_
```

```python
accuracy_score(y_test, pred)
```
