```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("../input/heart-attack-analysis-prediction-dataset/heart.csv")
dataset.head()
```

```python
x = dataset.drop("output", axis = 1)
y = dataset["output"]
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.get_params()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
model.fit(x_train,y_train);
```

```python
y_preds = model.predict(x_test)
y_preds
```

```python
model.score(x_test,y_test)
```

```python
for i in range (10,400,10):
  print(f"Traing the model with estimators - {i}")
  model = RandomForestClassifier(n_estimators=i)
  model.fit(x_train,y_train)
  print(f"Accuracy - {model.score(x_test,y_test)*100} %")
  print("")
```

```python
model = RandomForestClassifier(n_estimators=370)
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
