```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import plotly.express as px
import plotly.graph_objects as go
import unicodedata
import plotly.figure_factory as ff

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv(os.path.join(dirname, filename))
```

```python
columns = data.columns
```

```python
for column in columns:
    print(data[column].value_counts())
```

No se tendrán en cuenta los años con guerra cívil (Spanish Civil War (League Cancelled))

```python
index_league_cancelled = data[data['Goals'] == 'Spanish Civil War (League Cancelled)'].index
data.drop(index_league_cancelled, inplace=True)
#Normalize text
for column in columns:
    if column != 'Goals':
        data[column] = data[column].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8').str.strip()

data['Goals'] = data['Goals'].astype(int)
```

```python
data_winner_counts = data['Winner'].value_counts()
data_runnerup_counts = data['Runner-up'].value_counts()
data_thirdplace_counts = data['Third Place'].value_counts()
data_topscorer_counts = data['Top Scorer'].value_counts()
data_topscorer_team_counts = data["Top scorer's club(s)"].value_counts()
data_topscorer_sum = data.groupby(['Top Scorer'], as_index=False).agg('sum')
```

```python
fig1 = px.bar(data_winner_counts, color=data_winner_counts, title='Fig 1. Winners')
fig1.show()
```

```python
fig2 = px.bar(data_runnerup_counts, color=data_runnerup_counts, title='Fig 2. Runners-up')
fig2.show()
```

```python
fig3 = px.bar(data_thirdplace_counts, color=data_thirdplace_counts, title='Fig 3. Third Place')
fig3.show()
```

```python
fig4 = px.bar(data_topscorer_counts, color=data_topscorer_counts, title='Fig 4. Top Scorer')
fig4.show()
```

```python
fig5 = px.bar(data, x='Top Scorer', y='Goals', title='Fig 5. Top Scorer')
fig5.add_trace(go.Scatter(x=data_topscorer_sum['Top Scorer'], y=data_topscorer_sum['Goals'], mode='markers', name='Sum', marker_color=data['Goals']))
fig5.show()
```

```python
fig6 = px.bar(data_topscorer_team_counts, title='Fig 6. Top Scorer')
fig6.show()
```

```python
fig7 = px.box(data, x='Goals', points='all')
fig7.show()
```
