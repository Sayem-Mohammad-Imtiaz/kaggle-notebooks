```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
```

# Score of Different regions of the World


```python
plt.figure(figsize=(15,8))
sns.barplot(y='Regional indicator',x='Ladder score',data=df,palette="Blues_d",ci=0).set_title("Region-wise Score")
plt.show()
```

# Region-wise plot for Freedom vs Ladder score


```python
plt.figure(figsize=(15,8))
sns.lineplot(x="Ladder score",y="Freedom to make life choices",data=df,hue='Regional indicator')
plt.show()
```

# How Social Support impacts Score of the country?

```python
plt.figure(figsize=(15,8))
sns.lineplot(x="Ladder score",y="Social support",data=df)
plt.show()
```

# Higher the GDP per capita, Higher the score

```python
plt.figure(figsize=(15,8))
sns.jointplot(x="Ladder score",y="Logged GDP per capita",data=df,kind='hex')
plt.show()
```

# Perception of corruption of various regions of the world

```python
plt.figure(figsize=(15,8))
#sns.displot(data=df,y='Perceptions of corruption',x='Ladder score')
sns.barplot(data=df,x='Perceptions of corruption',y='Regional indicator',ci=0)
plt.show()
```
