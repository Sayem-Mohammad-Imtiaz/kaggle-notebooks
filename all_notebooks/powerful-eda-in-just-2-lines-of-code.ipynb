```python
import sweetviz
from IPython.display import Image
```

```python
train=pd.read_csv(r'/kaggle/input/hr-analytics-job-change-of-data-scientists/aug_train.csv')
test=pd.read_csv(r'/kaggle/input/hr-analytics-job-change-of-data-scientists/aug_test.csv')
```

```python
print("Size of train set: ", train.shape, "Size of test set: ",test.shape)
```

```python
train.head()
```

We now have 2 dataframes (train and test), and we would like to analyze the target value “target”.

```python
report = sweetviz.compare([train, "Train"], [test, "Test"], "target")
```

```python
report.show_html("Report.html")
```
