```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../input/manga-plots-and-demographics/manga_plot_demographic_wiki.csv", encoding="UTF-8")
df.shape
```

```python
df.head()
```

```python
pd.DataFrame(df.Demographic.unique()).values
```

```python
from wordcloud import WordCloud,STOPWORDS

plt.figure(figsize=(40,25))
subset1 = df[df['Demographic']=='Shōjo']
text = subset1.Plot.values
cloud1=WordCloud(stopwords=STOPWORDS,background_color='pink',colormap="Dark2",collocations=False,width=2500,height=1800).generate(" ".join(text))

plt.subplot(2, 2, 1)
plt.axis('off')
plt.title("Shōjo",fontsize=40)
plt.imshow(cloud1)

subset2 = df[df['Demographic']=='Shōnen']
text = subset2.Plot.values
cloud2=WordCloud(stopwords=STOPWORDS,background_color='pink',colormap="Dark2",collocations=False,width=2500,height=1800
                       ).generate(" ".join(text))
plt.subplot(2, 2, 2)
plt.axis('off')
plt.title("Shōnen",fontsize=40)
plt.imshow(cloud2)

subset3 = df[df['Demographic']=='Seinen']
text = subset3.Plot.values
cloud3=WordCloud(stopwords=STOPWORDS,background_color='pink',colormap="Dark2",collocations=False,width=2500,height=1800
                       ).generate(" ".join(text))
plt.subplot(2, 2, 3)
plt.axis('off')
plt.title("Seinen",fontsize=40)
plt.imshow(cloud3)

subset4 = df[df['Demographic']=='Josei']
text = subset4.Plot.values
cloud4=WordCloud(stopwords=STOPWORDS,background_color='pink',colormap="Dark2",collocations=False,width=2500,height=1800
                       ).generate(" ".join(text))
plt.subplot(2, 2, 4)
plt.axis('off')
plt.title("Josei",fontsize=40)
plt.imshow(cloud4)
```
