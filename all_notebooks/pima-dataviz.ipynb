```python
df = pd.read_csv("/kaggle/input/pima-indians-diabetes-database/diabetes.csv", sep=',')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#add a new column for a better visualisation
df['Overweight'] = [1 if x > 25 else 0 for x in df.BMI]
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

* s parameter: size of marker
* c parameter: color of marker. Might be used to see the class distribution accross the features. 
* alpha parameter: opacity of marker

```python
plt.scatter(df.Age, df.Insulin, s=50, cmap='viridis')
plt.colorbar()
plt.xlabel('Age')
plt.ylabel('Insulin')
plt.title('Relation between age and insulin levels')
plt.show()
```

## Scatter plot with subplot

```python
fig, axs = plt.subplots(1,2)
axs[0].scatter(df.Age, df.Glucose, s=50, cmap='viridis')
axs[0].set_title('Age versus Glucose')
axs[1].scatter(df.Age, df.BloodPressure)
axs[1].set_title('Age versus Blood Pressure')
plt.show()
```

## Histogram

It helps us interpret the distribution of variable. 

```python

fig, ax = plt.subplots()
ax.hist(df.BloodPressure, bins=10, histtype='bar')
ax.set_title('Histogram of Blood Pressure feature')
ax.set_xlabel('Blood Pressure')
ax.set_ylabel('Number of Observations')
plt.show()
```

We can give custom bins

```python
bins = [21, 31, 41, 51, 61, 71, 81]
fig, ax = plt.subplots()
ax.hist(df.Age, bins=bins, histtype='bar')
ax.set_title('Histogram of Age feature')
ax.set_xlabel('Age')
ax.set_ylabel('Number of Observations')
plt.show()
```

## Bar plot for categorical variables 

```python
fig, ax = plt.subplots()
ax.bar(['no diabete', 'diabete'], df.groupby('Outcome')['Age'].mean(), color=['red', 'blue'])
ax.set_xlabel('Outcome')
ax.set_ylabel('Average age')
fig.savefig('AgeVersusOutcome.png')
plt.show()
```

# Seaborn

## Count Plot

**value counts for single categorical variable**

```python
sns.set_palette("RdBu") #set a color palette
#countplot create a histogram across a categorical variable
sns.countplot(x='Outcome', data=df)
plt.show()
```

**value counts for two categorical variable**

```python
sns.countplot(x='Outcome', hue='Overweight', data=df)
plt.show()
```

for more, visit https://seaborn.pydata.org/generated/seaborn.countplot.html

```python
#catplot 
sns.catplot(x="Age", aspect=3, data=df, kind='count')
plt.show()
```

### set title

```python
g = sns.catplot(x="Age", aspect=3, data=df, kind='count')
g.fig.suptitle('Age Counts')
plt.show()
```

### Rotate xticks

```python
g = sns.catplot(x="Age", aspect=3, data=df, kind='count')
plt.xticks(rotation=30)
plt.show()
```

## Scatter Plot

Same as plt.scatter

```python
sns.scatterplot(x='Age', y='Insulin', data=df, hue='Outcome')
plt.show()
```

relplot is a more general way

```python
sns.relplot(x='Age', y='Insulin', data=df, hue='Outcome', kind='scatter')
plt.show()
```
