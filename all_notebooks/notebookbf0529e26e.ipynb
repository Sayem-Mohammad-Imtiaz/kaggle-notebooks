```python
# col_names = ["Permit Type", "Permit Creation Date", "Block", "Lot", "Existing Construction Type", "Supervisor District", "Zipcode"]

df = pd.read_csv("/kaggle/input/building-permit-applications-data/Building_Permits.csv", header=None, names = [""])

df
```

```python
# col_names = [1, 3, 4, 5, 15, 17, 20, 21, 23, 24, 25, 26, 27, 28, 29, 30, 33, 37, 40]
col_names = [1, 20, 25, 26, 40]
df[col_names]
```

```python
# converting permit type to int
df = df.iloc[1:]
df[1]=df[1].apply(int)
df[1][2]


```

```python
# pd.to_datetime(df[3]).value
# df[3].apply(lambda x: x.value)
# df[3].apply(pd.Timestamp).strftime("%Y-%m-%d")
# pd.to_datetime(df[3], format='%m/%d/%Y')
df[3] = df[3].astype(str)
df[3] = df[3].str.replace('\D', '').astype(int)
df[3]
```

```python
feature_cols = [1, 3, 4, 5, 15, 17, 20, 21, 23, 24, 25, 26, 27, 28, 29, 30, 33, 37, 40]
X = df[feature_cols] # Features
y = df[13]
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
