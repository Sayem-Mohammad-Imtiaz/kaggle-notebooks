```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)


proc_data = pd.read_csv("../input/procurement-notices.csv",parse_dates=True, infer_datetime_format=True)
#import os
#print(os.listdir("../input"))


proc_data.dtypes
```

```python
#proc_data.head(10)
cols = ['id', 'url', 'notice_type', 'publication_date', 'project_id', 'bid_description', 'procurement_type', 'deadline_date', 'country_code', 'country_name', 'major_sector']
proc_data.columns = cols
proc_data.dtypes
```

```python
proc_data.deadline_date = pd.to_datetime(proc_data['deadline_date'])
```

```python
#number of calls currently out
# cells with NaN deadline date are currently out, not selected here
date_null = pd.isna(proc_data['deadline_date'])
#proc_data[date_null]
current_calls = proc_data[proc_data['deadline_date'] > now] 
#current_calls = proc_data[proc_data['deadline_date'] > yesterday]
current_calls.head(10)
```

```python
date_null = pd.isna(proc_data['deadline_date'])
df_null_date = proc_data[date_null]
df_null_date.head(10)
```

```python
total_current_calls = pd.concat([current_calls, df_null_date])
total_current_calls.count()
```

```python
deadlines = total_current_calls['deadline_date'].value_counts()
deadlines[:len(deadlines)].plot(kind='barh')
```

```python
## day 2
#n_countries = proc_data['country_name'].value_counts()
#print(n_countries)
df_countries = proc_data['project_id'].groupby(proc_data['country_name']).count()
#print(df_countries)
```

```python
df_countries = df_countries.reset_index()
df_countries.rename(columns ={'project_id':'count'}, inplace = True)
df_countries = df_countries.sort_values('count', ascending = False)
df_countries.reset_index(drop=True, inplace = True)
```

```python
print(df_countries)
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

# these two lines are what allow your code to show up in a notebook!
from plotly.offline import init_notebook_mode, iplot
init_notebook_mode()
```

```python
# specify what we want our map to look like
data = dict(
        type='choropleth',
        autocolorscale = False,
        locations = df_countries['country_name'],
        z = df_countries['count'],
        locationmode = 'country names',
        colorscale=[[0, 'rgb(224,255,255)'],
            [0.01, 'rgb(166,206,227)'], [0.02, 'rgb(31,120,180)'],
            [0.03, 'rgb(178,223,138)'], [0.05, 'rgb(51,160,44)'],
            [0.10, 'rgb(251,154,153)'], [0.20, 'rgb(255,255,0)'],
            [1, 'rgb(227,26,28)']],    
reversescale = False
       ) 

# chart information
#layout = dict(title='Number of films in the TMDB database',
#geo = dict(showframe = True, projection={'type':'Mercator'}))
layout = dict(
        title = 'Number of Procurements by Country',
        geo = dict(
            showframe = True, projection={'type':'mercator'})
)
             
choromap = go.Figure(data = [data], layout = layout)
iplot(choromap, validate=False)
# actually show our figure
#fig = dict( data=data, layout=layout )
#iplot( fig, filename='d3-cloropleth-map' )
```
