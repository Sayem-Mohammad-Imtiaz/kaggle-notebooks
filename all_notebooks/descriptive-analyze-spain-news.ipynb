# Step 1 - EDA and extract the 50 most used words in title and content

```python
df = pd.read_csv("../input/noticias-laraznpblico/data_larazon_publico_v2.csv")

df.info()
```

```python
with open('../input/noticias-laraznpblico/data_larazon_publico_v2.csv', newline='', encoding="utf8") as File:  
    reader = csv.reader(File, delimiter=",")
    
    #Get words in titles
    w_title = list()
    w_content = list()
    
    for row in reader:
        title = row[3]
        
        words_t = title.split(" ")
        
        for w in words_t:
            if len(w)>4:
                w_title.append(w.lower())
    
        #Get words in content
        content = row[2]
        
        words_c = content.split(" ")
        
        for w in words_c:
            if len(w)>4:
                w_content.append(w.lower())
```

```python
c = Counter(w_title)
data_t = {}
print ("Ranking | Word | Frequency")
for i in range(0,50):
    title_wf = c.most_common()[i][0]
    title_wfc = c.most_common()[i][1]
    print(f"   {i}   :  {title_wf}  :    {title_wfc}   ")
    data_t[title_wf] = title_wfc
    if i == 50:
        break
```

```python
c = Counter(w_content)
data_c = {}
print ("Ranking | Word | Frequency")
for i in range(0,50):
    content_wf = c.most_common()[i][0]
    content_wfc = c.most_common()[i][1]
    print(f"   {i}   :  {content_wf}  :    {content_wfc}   ")
    data_c[content_wf] = content_wfc
    if i == 50:
        break
```

# Step 2 - Visualization

Most frequent words in titles

```python
df = pd.DataFrame.from_dict(data_t, orient='index')
muestra = df[:20].copy()
muestra.plot(kind = "bar")
```

Most frequent words in news content

```python
df = pd.DataFrame.from_dict(data_c, orient='index')
muestra = df[:20].copy()
muestra.plot(kind = "bar")
```
