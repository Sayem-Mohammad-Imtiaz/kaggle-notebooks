### Questão 1 - Item B - Tabela de frequência

Construa uma tabela de frequência para cada uma das **variáveis qualitativas** que você escolheu (caso não tenha escolhido nenhuma, deixe esta questão em branco). Uma dica: a função *value_counts()* do Pandas pode ser muito útil. =)


## Resposta - Questão 1 - Item B - Tabela de frequência

```python
modelos = df['aeronave_tipo_veiculo'].value_counts().to_frame(name='Count')
modelos.head()
```

```python
tipo_motor = df['aeronave_motor_tipo'].value_counts().to_frame(name='Count')
tipo_motor
```

```python
pais_fab = df['aeronave_pais_fabricante'].value_counts().to_frame(name='Count')
pais_fab.head()
```

```python
pais_reg = df['aeronave_pais_registro'].value_counts().to_frame(name='Count')
pais_reg.head()
```

```python
tipo_oper = df['aeronave_tipo_operacao'].value_counts().to_frame(name='Count')
tipo_oper.head()
```

```python
dano = df['aeronave_nivel_dano'].value_counts().to_frame(name='Count')
dano.head()
```

### Questão 1 - Item C - Representação Gráfica
Para cada uma das variáveis, produza um ou mais gráficos, usando matplotlib, que descreva seu comportamento / caracteristica. Lembre-se que estes gráficos precisam ser compatíveis com a classificação da variável. 

## Resposta - Questão 1 - Item C - Representação Gráfica

```python
import matplotlib.pyplot as plt
```

```python
#Creating a figure with some fig size
fig, ax = plt.subplots(figsize = (15,10))

models = list(modelos['Count'])
plt.bar(list(modelos.index), list(modelos['Count']))
plt.title('Quantidade de aeronaves de cada modelo', fontsize=15)
plt.ylabel('Quantidade', fontsize=15)
plt.xticks(list(modelos.index), list(modelos.index), rotation='horizontal')
plt.xlabel('Modelos', fontsize=15)

for index,data in enumerate(models):
    plt.text(x=index - (len(str(data))/18) , y =data + 50 , s=f"{data}" , fontdict=dict(fontsize=15))

plt.show()
```

```python
#Creating a figure with some fig size
fig, ax = plt.subplots(figsize = (15,10))

motors = list(tipo_motor['Count'])
plt.bar(list(tipo_motor.index), list(tipo_motor['Count']))
plt.title('Quantidade de aeronaves de cada tipo de motor', fontsize=15)
plt.ylabel('Quantidade', fontsize=15)
plt.xticks(list(tipo_motor.index), list(tipo_motor.index), rotation='horizontal')
plt.xlabel('Tipo de motor', fontsize=15)

for index,data in enumerate(motors):
    plt.text(x=index - (len(str(data))/28) , y =data + 50 , s=f"{data}" , fontdict=dict(fontsize=15))

plt.show()
```

```python
#Creating a figure with some fig size
fig, ax = plt.subplots(figsize = (15,10))

countrys = list(pais_fab['Count'])
plt.bar(list(pais_fab.index), list(pais_fab['Count']))
plt.title('Quantidade de aeronaves fabricadas em cada país', fontsize=15)
plt.ylabel('Quantidade', fontsize=15)
plt.xticks(list(pais_fab.index), list(pais_fab.index), rotation='vertical')
plt.xlabel('País de Fabricação', fontsize=15)

for index,data in enumerate(countrys):
    plt.text(x=index - (len(str(data))/10) , y =data + 40 , s=f"{data}" , fontdict=dict(fontsize=15))

plt.show()
```

```python
#Creating a figure with some fig size
fig, ax = plt.subplots(figsize = (15,10))

registry = list(pais_reg['Count'])
plt.bar(list(pais_reg.index), list(pais_reg['Count']))
plt.title('Quantidade de aeronaves registradas em cada país', fontsize=15)
plt.ylabel('Quantidade', fontsize=15)
plt.xticks(list(pais_reg.index), list(pais_reg.index), rotation='vertical')
plt.xlabel('País de Registro', fontsize=15)

for index,data in enumerate(registry):
    plt.text(x=index - (len(str(data))/8) , y =data + 40 , s=f"{data}" , fontdict=dict(fontsize=15))

plt.show()
```

```python
#Creating a figure with some fig size
fig, ax = plt.subplots(figsize = (15,15))

operations = list(tipo_oper['Count'])
plt.bar(list(tipo_oper.index), list(tipo_oper['Count']))
plt.title('Quantidade de aeronaves por tipos de operações', fontsize=15)
plt.ylabel('Quantidade', fontsize=15)
plt.xticks(list(tipo_oper.index), list(tipo_oper.index), rotation='vertical')
plt.xlabel('Tipos de operações', fontsize=15)

for index,data in enumerate(operations):
    plt.text(x=index - (len(str(data))/15) , y =data + 30 , s=f"{data}" , fontdict=dict(fontsize=15))

plt.show()
```

```python
#Creating a figure with some fig size
fig, ax = plt.subplots(figsize = (8,8))

damage = list(dano['Count'])
plt.bar(list(dano.index), list(dano['Count']))
plt.title('Quantidade de aeronaves por niveis de danos',fontsize=15)
plt.ylabel('Quantidade', fontsize=15)
plt.xticks(list(dano.index), list(dano.index), rotation='vertical')
plt.xlabel('Nível de dano', fontsize=15)

for index,data in enumerate(damage):
    plt.text(x=index - (len(str(data))/15) , y =data + 20 , s=f"{data}" , fontdict=dict(fontsize=15))

plt.show()
```

```python
#Creating a figure with some fig size
fig, ax = plt.subplots(figsize = (15,7))

fabricacao = df['aeronave_ano_fabricacao'].value_counts().to_frame(name='Count').sort_index()

year = list(fabricacao.index)
count = list(fabricacao['Count'])

#Removendo outlier
count.pop(year.index(0))
year.remove(0)

plt.plot(year, count)
plt.title('Quantidade de aeronaves produzidas por ano', fontsize=15)
plt.ylabel('Quantidade', fontsize=15)
plt.xlabel('Anos', fontsize=15)

plt.show()
```
