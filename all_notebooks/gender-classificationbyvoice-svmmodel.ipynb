```python
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
from sklearn.svm import SVC
```

```python
data=pd.read_csv('../input/voicegender/voice.csv')  #importing the dataset in pandas dataframe
data  #visualize the dataframe
```

```python
data.describe()    #shows the datastats
```

```python
def male_female_label(gender):    #changes the labels  (male,female) to label markers i.e male=0 female=1
    if gender[0]=='m':
        return 0
    else:
        return 1
    
```

```python
data['label']=data['label'].apply(male_female_label)  #applies the label
```

One can also use the Label Encoder from sklearn.preprocessing

```python
data   
```

```python
profile=ProfileReport(data)   #data_analysis
profile
```

```python
X=data.drop('label',axis=1)
y=data['label']            #separating the features and output


scaler=MinMaxScaler()       #scaling the values between 0 to 1
X_scaled=scaler.fit_transform(X)

X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,train_size=0.7)    #spiliting the data into train set and test set..with training set to be 30% of the total data
```

```python
model=SVC(C=1,gamma=1,kernel='rbf')
model.fit(X_train,y_train)                 #defining the support vector model and training


y_pred=model.predict(X_test)               #predicting from test data

acc=model.score(X_test,y_test)             #calculating the accuracy and score of the model
score=f1_score(y_test,y_pred)



print(f"Support Vector Machine:\nAccuracy: {acc}\nF1 Score: {score}\n")
```
