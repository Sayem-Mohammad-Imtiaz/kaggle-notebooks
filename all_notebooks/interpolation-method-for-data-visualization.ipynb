```python
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d
```

```python
data = pd.read_csv(r'/kaggle/input/videogamesales/vgsales.csv')
data.describe()
```

```python
display(data)
```

```python
def plot(x,y):
    f = interp1d(x, y)
    f2 = interp1d(x, y, kind='cubic')

    xnew = np.linspace(1, len(x) , endpoint=True)
    plt.figure(figsize=(20,5))
    plt.plot(x, y, 'o', xnew, f(xnew), '-', xnew, f2(xnew), '--')
    plt.legend(['data', 'linear', 'cubic'], loc='best')
    plt.show()
```

```python

data.sort_values(by='NA_Sales',ascending=True).head()
x = data['Rank'].tolist()
y = data['NA_Sales'].tolist()
plot(x,y)
```

```python
#Average sales in every year of all the games
df = data.groupby('Year').mean()
df
x = df.index.tolist()
y = df['Global_Sales']
print(x)
f = interp1d(x, y)
f2 = interp1d(x, y, kind='cubic')

xnew = np.linspace(x[0], x[len(x)-1] , endpoint=True)
plt.figure(figsize=(20,5))
plt.plot(x, y, 'o', xnew, f(xnew), '-', xnew, f2(xnew), '--')
plt.legend(['data', 'linear', 'cubic'], loc='best')
plt.show()
```

```python
#Average sales in every year of all the Wii platform
df = data[data['Platform']=="Wii"]
df = df.groupby('Year').mean()

x = df.index.tolist()
y = df['Global_Sales']
print(x)
f = interp1d(x, y)
f2 = interp1d(x, y, kind='cubic')

xnew = np.linspace(x[0], x[len(x)-1] , endpoint=True)
plt.figure(figsize=(20,5))
plt.plot(x, y, 'o', xnew, f(xnew), '-', xnew, f2(xnew), '--')
plt.legend(['data', 'linear', 'cubic'], loc='best')
plt.show()
```

```python
df = data[data['Publisher']=="Nintendo"]
df = df.groupby('Year').mean()
display(df.head())


from scipy.interpolate import Rbf
x = df['EU_Sales'].to_numpy()
y = df['NA_Sales'].to_numpy()
z = df['Rank'].to_numpy()
edges = np.linspace(0,5, 101)
centers = edges[:-1] + np.diff(edges[:2])[0] / 2.
XI, YI = np.meshgrid(centers, centers)


# use RBF
rbf = Rbf(x, y, z, epsilon=2)
ZI = rbf(XI, YI)
# plot the result
plt.subplot(1, 1, 1)
X_edges, Y_edges = np.meshgrid(edges, edges)
lims = dict(cmap='RdBu_r', vmin=min(z), vmax=max(z))
plt.pcolormesh(X_edges, Y_edges, ZI, shading='flat', **lims)
plt.scatter(x, y, 100, z, edgecolor='w', lw=0.1, **lims)
plt.title('RBF interpolation - multiquadrics')
plt.xlabel("EU SALES")
plt.ylabel('NA SALES')
plt.colorbar()
```
