```python
X = df.drop("output" , axis =1)
y = df['output']
```

# Feature Scaling

```python
#Standardising the features
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() #It will make the features to have mean = 0 and standard deviation = 1
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.2 , random_state=0, stratify=df.output)
```

# Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

models = [LogisticRegression, DecisionTreeClassifier, KNeighborsClassifier]
model_accuracy = {}
for model in models:
    model = model()
    model.fit(X_train , y_train)
    model_accuracy[model.__class__.__name__] = model.score(X_test , y_test)
```

```python
model_accuracy
```

```python
# Lets take Logistic Regression and tune it to get better results
#This algorithm has C parameter that will affect the models performane
LogisticRegression_C = [1e-3, 1e-2, 1e-1, 1, 10]
model_accuracy_with_C = {}
for c in LogisticRegression_C:
    model = LogisticRegression(C=c)
    model.fit(X_train , y_train)
    y_pred = model.predict(X_test)
    score = accuracy_score(y_pred , y_test)
    model_accuracy_with_C[str(c)] = accuracy_score
    print(f"The accuracy score for {model.__class__.__name__} and {c} is {score}")
```

we see as we increase the value of C the model performance is increasing and at one point it is plateau

```python
#Training the LogisticRegression model with parameter C=10

final_model = LogisticRegression(C=10)
final_model.fit(X_train ,y_train)
y_pred = final_model.predict(X_test)
print(f"Accuracy is: {(accuracy_score(y_pred , y_test)*100):.2f}")
```
