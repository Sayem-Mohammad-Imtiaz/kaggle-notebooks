```python
import pandas as pd

df=pd.read_csv('../input/youtube-video-statistics/TODAY.csv')
df
```

```python
#Посчитать статистические показатели
import statistics as st
import numpy as np

print('Медиана:', st.median(df['likes']))
print('Мода:',st.mode(df['likes']))
print('Среднее:',st.mean(df['likes']))
print('Минимум:',np.min(df['likes']))
print('Максимум:',np.max(df['likes']))
```

```python
#Создать новый столбец, в котором разместить разницу между максимумом и минимумом
df['Разница']=df.max(axis = 1) - df.min(axis = 1)
df
```

```python
#Разбить данные на несколько блоков (не менее 3, приблизительно равных по размеру) по некоторому критерию 
block1=df[(df['likes'])<=50]
print('Менее 50 лайков:', len(block1.index))
block2=df[(df['likes']>50)&(df['likes']<=250)] 
print('50-250 лайков:',len(block2.index))
block3=df[(df['likes']>250)&(df['likes']<=5000)]
print('250-5000 лайков:',len(block3.index))
block4=df[(df['likes']>5000)]
print('Более 5000 лайков:', len(block4.index))
```

```python
#Привести круговую диаграмму
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

vals = [len(block1.index),len(block2.index), len(block3.index), len(block4.index)]
labels = ["Менее 50 лайков", "50-250 лайков", "250-5000 лайков", "Более 5000 лайков"]
fig, ax = plt.subplots()
ax.pie(vals, labels=labels)
ax.axis("equal")

ax.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
plt.show()
```

```python
#В зависимости от особенностей набора данных для каждого блока построить график
plt.plot(block1['likes'], block1['views'])
plt.show()
plt.hist(block1['likes'], color='blue', alpha=0.7, rwidth=0.85)
plt.show()


plt.plot(block2['likes'], block2['views'])
plt.show()
plt.hist(block2['likes'], color='blue', alpha=0.7, rwidth=0.85)
plt.show()

plt.plot(block3['likes'], block3['views'])
plt.show()
plt.hist(block3['likes'], color='blue', alpha=0.7, rwidth=0.85)
plt.show()

plt.plot(block4['likes'], block4['views'])
plt.show()
plt.hist(block4['likes'], color='blue', alpha=0.7, rwidth=0.85)
plt.show()
```

```python
#Сгруппировать данные (GroupBy) по некоторому признаку и сохранить результаты в новые таблицы. 
#Для каждой новой таблицы провести сортировку по сложному ключу, состоящему из нескольких признаков. 
#Для каждой таблицы своя сортировка.

#Достать из даты год, чтобы сгруппировать по нему
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
#df['year']

df_group = dict(tuple(df.groupby('year')))

#Дата сортируется по возрастанию, просмотры по убыванию
df1=df_group[2017].sort_values(by=['date','views'], ascending=[True, False])
print(df1)

df2=df_group[2018].sort_values(by=['date','views'], ascending=[True, False])
print(df2)

df3=df_group[2019].sort_values(by=['date','views'], ascending=[True, False])
print(df3)

df4=df_group[2020].sort_values(by=['date','views'], ascending=[True, False])
print(df4)
```

```python
#Отобразить распределение некоторого параметра по дням недели (в каждой группе), по месяцам (суммарно).
import calendar

# Количество просмотров в день недели
kol_day = df1[(df1['views'])>1].groupby(df1['date'].dt.dayofweek).count().date
# Замена целых чисел дня названиями дней
kol_day.index=[calendar.day_name[x] for x in range(0,7)]
kol_day.plot(kind='bar', color='blue')
plt.title('Распределение Views по дням недели')
plt.xlabel('Day of week')
plt.ylabel('Count')

```
