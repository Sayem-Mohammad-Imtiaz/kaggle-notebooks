```python
df = pd.read_csv("/kaggle/input/tokyo-weather-data/tokyo_data.csv")
# Read the first five rows
df.head(5)
```

```python
# Read the summary of the data frame
df.describe()
```

```python
# Read the data type of each column
df.dtypes

# Change the name of each column to a better understanding way
df2 = df.rename(columns = {'daylight_hours(hours)':'daylight', 'high(��)':'high', 'low(��)': 'low', 'y/m/d': 'date', 'max_precipitation_10min(mm)': 'max_precipitation_10min'})
```

```python
# Change the daylight to numeric form from object to float/ string
df2.daylight = pd.to_numeric(df2['daylight'], errors = 'coerce')
df2.high = pd.to_numeric(df2['high'], errors = 'coerce')
df2.low = pd.to_numeric(df2['low'], errors = 'coerce')
df2.max_precipitation_10min = pd.to_numeric(df2['max_precipitation_10min'], errors = 'coerce')
df2.date = df2.date.astype('string')

# Read the data type of each column again
df2.dtypes
```

```python
# Plot the linear regression line for the last 300 rows
df3 = df2.tail(300)
sns.lmplot(x='high', y='daylight', data=df3);
```
