```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('../input/fantasypremierleague1920updated/players_1920_fin_updated_role.csv')
print(df.shape)
df.head()
```

```python
role_dict={'GK':'Goal Keeper', 'DEF':'Defender',
           'MID': 'Midfielder', 'FWD':'Forward'}
```

```python
x=[df.total_points[df.was_home==True].sum(),
   df.total_points[df.was_home==False].sum()]
labels=['Home', 'Away']
explode=[0.02,0.02]
plt.figure(figsize=(4,4))
plt.title('Points earned Home Vs Away')
plt.pie(x=x, labels=labels, autopct='%1.1f%%', 
        startangle=90, explode=explode)
plt.show()
```

```python
labels=['Home', 'Away']
explode=[0.02,0.02]
plt.figure(figsize=(10,4))
plt.suptitle('Points earned Home Vs Away', fontsize=14)

for i in range(0,4):
    role=list(role_dict.keys())[i]
    temp_df=df[df.role==role]
    
    x=[temp_df.total_points[df.was_home==True].sum(),
       temp_df.total_points[df.was_home==False].sum()]
    
    plt.subplot(1,4,i+1)
    plt.title(role_dict[role], fontsize=8)
    plt.pie(x=x, labels=labels, autopct='%1.1f%%', 
            startangle=90, explode=explode)
plt.show()
```

```python
df_temp=df.pivot_table(values='total_points', aggfunc='sum', 
               index='team', columns='was_home').reset_index()
df_temp['Total']=df_temp[1]+df_temp[0]
df_temp["Home%"]=round(df_temp[1]*100/df_temp.Total,2)
df_temp.sort_values(by='Total', ascending=False, inplace=True)
sns.scatterplot(data=df_temp, y='Total', x='Home%')
df_temp.style.background_gradient(cmap='coolwarm',subset=['Home%'])
```

```python
df_temp=df.pivot_table(values='total_points', aggfunc='sum', 
               index=['team','role'], columns='was_home').reset_index()
df_temp['Total']=df_temp[1]+df_temp[0]
df_temp["Home%"]=round(df_temp[1]*100/df_temp.Total,2)
sns.scatterplot(data=df_temp, y='Total', x='Home%', hue='role')
df_temp.sort_values(by='Home%', ascending=False, inplace=True)
df_temp.style.background_gradient(cmap='coolwarm',subset=['Home%'])
```
