```python
df['Inactive_flag'] = df['Inactive_flag'].fillna('Active').replace('wi','InActive')
```

```python
plt.figure(figsize=(18,6))
fig = px.bar(x=df.Country.value_counts().index, y=df.Country.value_counts(), title='Country Distribution')
fig.show()
```

```python
df['Title'].value_counts(dropna=False)
```

column Title has 5435 nan values.

```python
fig = px.bar(x=df['Title'].value_counts().index, y=df['Title'].value_counts(), title='Chess Title Distribution')
fig.show()
```

Woman FIDE master = WFM has the highest count.

Grand Master = GM is the highest title in chess game has only 37 count.

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=df['Standard_Rating'],
                           xbins=dict(
                          start=1800,
                          end=2675,
                          size=6),
                          opacity=1))

fig.update_layout(title_text='Standard_Rating Distribution',
                 xaxis_title='Standard_Rating',
                 yaxis_title='Count',
                 xaxis ={'showgrid':False},
                 yaxis ={'showgrid':False},
                 bargap=0.05,
                 template='plotly_dark')
fig.show()
```

```python

fig = px.histogram(df,x=df['Standard_Rating'], color='Inactive_flag')
fig.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=df['Rapid_rating'],
                           xbins=dict(
                          start=1220,
                          end=2650,
                          size=7),
                          opacity=1))

fig.update_layout(title_text='Rapid_rating Distribution',
                 xaxis_title='Rapid_rating',
                 yaxis_title='Count',
                 xaxis ={'showgrid':False},
                 yaxis ={'showgrid':False},
                 bargap=0.05,
                 template='seaborn')
fig.show()
```

```python
fig = px.pie(values=df['Inactive_flag'].value_counts(), names=df['Inactive_flag'].value_counts().index, title='Active VS InActive Players',
            color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```

```python
gm = df[df['Title']=='GM']
gm_stand = gm.sort_values(by=['Standard_Rating'],ascending=False).loc[:,['Name','Country','Standard_Rating']].head(10)
gm_stand.style.background_gradient(cmap='YlGnBu', low=0, high=0, axis=0, subset=None)
```

```python
gm_rapid = gm.sort_values(by='Rapid_rating', ascending=False).loc[:,['Name','Country','Rapid_rating']].head(10)
gm_rapid.style.background_gradient(cmap='YlGnBu', low=0, high=0, axis=0, subset=None)
```

```python
gm_blitz = gm.sort_values(by='Blitz_rating', ascending=False).loc[:,['Name','Country','Blitz_rating']].head(10)
gm_blitz.style.background_gradient(cmap='YlGnBu', low=0, high=0, axis=0, subset=None)
```

Polgar, Judit from Hungary has highest Standard, Rapid and Blitz rating.
