```python
insurance = pd.read_csv('../input/insurance/insurance.csv')
```

```python
insurance.head(1000).plot.scatter(x = 'age' , y = 'charges' )
```

```python
insurance.plot.hexbin(y = 'age' , x = 'charges',gridsize=15 )
```

We can see charges increase as the age increase
