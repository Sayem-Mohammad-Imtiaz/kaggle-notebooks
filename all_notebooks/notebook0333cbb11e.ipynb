```python
#read bike_buyers csv file
df = pd.read_csv('/kaggle/input/bike-buyers/bike_buyers.csv')
```

```python
#getting information about data about non-null count and dtype
df.info()
```

```python
#discovery of data to watch the unique, top and freq amounts
df.describe(include=['O'])
```

```python
#mapping contents to numeric classification
df["Marital Status"] = df["Marital Status"].map({'Married' : '1', 'Single' : '0'})
df["Gender"] = df["Gender"].map({'Female' : '1', 'Male' : '0'})
df["Education"] = df["Education"].map({'Bachelors':'4', 'Partial College':'3', 'High School':'2',
'Partial High School':'1', 'Graduate Degree':'5'})
df["Home Owner"] = df["Home Owner"].map({'Yes' : '1', 'No' : '0'})
df["Purchased Bike"] = df["Purchased Bike"].map({'Yes' : '1', 'No' : '0'})
df["Region"] = df["Region"].map({'Europe' : '2', 'Pacific' : '1','North America' : 0})
df["Occupation"] = df["Occupation"].map({'Management':'4', 'Manual':'3', 'Professional':'2',
'Clerical':'1', 'Skilled Manual':'5'})
df["Commute Distance"] = df["Commute Distance"].map({'0-1 Miles': '1','1-2 Miles': '2','2-5 Miles': '5','5-10 Miles': '10'})

```

```python
#change from float to int
pd.options.display.float_format = '{:,.0f}'.format
```

```python
df.head()
```

```python
#drop missing values
df2 = df.dropna()
```

```python
#changing to numeric
df2.apply(pd.to_numeric, errors='ignore')
```

**VISUALIZATION**

```python
marital_bike = df2.groupby(['Marital Status'], as_index = False)['Purchased Bike'].count()
marital_bike.plot(kind = 'bar' , x = 'Marital Status' , y = 'Purchased Bike', figsize = (6 , 6) , color = 'red' , grid = True , width = 0.2 , rot = 0)
plt.tight_layout()
plt.show()

Gender_bike = df2.groupby(['Gender'], as_index = False)['Purchased Bike'].count()
Gender_bike.plot(kind = 'bar' , x = 'Gender' , y = 'Purchased Bike', figsize = (6 , 6) , color = 'blue' , grid = True , width = 0.2 , rot = 0)
plt.tight_layout()
plt.show()

Income_bike = df2.groupby(['Income'], as_index = False)['Purchased Bike'].count()
Income_bike.plot(kind = 'bar' , x = 'Income' , y = 'Purchased Bike', figsize = (15 , 6) , color = 'black' , grid = True , width = 0.2 , rot = 0)
plt.tight_layout()
plt.show()

Children_bike = df2.groupby(['Children'], as_index = False)['Purchased Bike'].count()
Children_bike.plot(kind = 'bar' , x = 'Children' , y = 'Purchased Bike', figsize = (15 , 6) , color = 'yellow' , grid = True , width = 0.2 , rot = 0)
plt.tight_layout()
plt.show()

Education_bike = df2.groupby(['Education'], as_index = False)['Purchased Bike'].count()
Education_bike.plot(kind = 'bar' , x = 'Education' , y = 'Purchased Bike', figsize = (15 , 6) , color = 'green' , grid = True , width = 0.2 , rot = 0)
plt.tight_layout()
plt.show()

Occupation_bike = df2.groupby(['Occupation'], as_index = False)['Purchased Bike'].count()
Occupation_bike.plot(kind = 'bar' , x = 'Occupation' , y = 'Purchased Bike', figsize = (15 , 6) , color = 'gray' , grid = True , width = 0.2 , rot = 0)
plt.tight_layout()
plt.show()

Cars_bike = df2.groupby(['Cars'], as_index = False)['Purchased Bike'].count()
Cars_bike.plot(kind = 'bar' , x = 'Cars' , y = 'Purchased Bike', figsize = (15 , 6) , color = 'pink' , grid = True , width = 0.2 , rot = 0)
plt.tight_layout()
plt.show()

Commute_bike = df2.groupby(['Commute Distance'], as_index = False)['Purchased Bike'].count()
Commute_bike.plot(kind = 'bar' , x = 'Commute Distance' , y = 'Purchased Bike', figsize = (15 , 6) , color = 'red' , grid = True , width = 0.2 , rot = 0)
plt.tight_layout()
plt.show()

Region_bike = df2.groupby(['Region'], as_index = False)['Purchased Bike'].count()
Region_bike.plot(kind = 'bar' , x = 'Region' , y = 'Purchased Bike', figsize = (15 , 6) , color = 'black' , grid = True , width = 0.2 , rot = 0)
plt.tight_layout()
plt.show()

Age_bike = df2.groupby(['Age'], as_index = False)['Purchased Bike'].count()
Age_bike.plot(kind = 'bar' , x = 'Age' , y = 'Purchased Bike', figsize = (30 , 6) , color = 'green' , grid = True , width = 0.2 , rot = 0)
plt.tight_layout()
plt.show()
```
