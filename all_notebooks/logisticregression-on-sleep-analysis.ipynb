# Why Logistic Regression Fails here!

```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler, RobustScaler
df=pd.read_csv("../input/sleep-pattern/Sleep_Analysis.csv")
df.head()
```

# Label encoding

```python
le=LabelEncoder()
df['Gender']=le.fit_transform(df['Gender'])
df['meals/day']=le.fit_transform(df['meals/day'])
df['physical illness']=le.fit_transform(df['physical illness'])
df['screen time']=le.fit_transform(df['screen time'])
df['bluelight filter']=le.fit_transform(df['bluelight filter'])
df['sleep direction']=le.fit_transform(df['sleep direction'])
df['exercise']=le.fit_transform(df['exercise'])
df['smoke/drink']=le.fit_transform(df['smoke/drink'])
df['beverage']=le.fit_transform(df['beverage'])
print("successful")
```

```python
df.head()
```

```python
X=df.drop(['sleep time'],axis=1)
Y=df['sleep time']
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression().fit(X_train,y_train)
r_sq = model.score(X_test, y_test)
print('coefficient of determination:', r_sq)
```
