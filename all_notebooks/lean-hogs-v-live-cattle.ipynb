```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# Reading in data
liveCattle = pd.read_csv('../input/commodity-futures-price-history/Commodity Data/Live Cattle.csv')
leanHogs = pd.read_csv('../input/commodity-futures-price-history/Commodity Data/Lean Hogs.csv')
```

```python
# Only using date and closing price data, removing all other columns
col2drop = ['Open', 'High', 'Low', 'Adj Close', 'Volume']

liveCattle.drop(columns = col2drop, axis = 0, inplace = True)
liveCattle.columns = ['date', 'cattleClose']

leanHogs.drop(columns = col2drop, axis = 0, inplace = True)
leanHogs.columns = ['date', 'hogsClose']
```

```python
# Join dataframes, correct type of date, drop NA values
df = pd.merge(liveCattle, leanHogs, on = 'date')
df['date'] = pd.to_datetime(df['date'])
df.dropna()
```

```python
# quick look at the data
df.plot(x = 'date')
```

```python
# Check correlation
df.corr()
```

```python
# Calculate spread of prices
df['spread'] = df['cattleClose']-df['hogsClose']
```

```python
df.plot(x = 'date', y = 'spread')
```
