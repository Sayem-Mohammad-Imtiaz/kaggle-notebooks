```python
import pandas as pd
import numpy as np
print('Priyatama is ready!')
```

```python
data = pd.read_csv('../input/india-census-dataset/India Census 2011 Data.csv')
data.head(2)
```

```python
data.dtypes
```

```python
#Q. 1) Hide the index.
data.style.hide_index()
```

```python
#Q. 2) Set caption on dataframe.
data.style.set_caption('India Census 2011')
```

```python
#Q. 3) Show the records related with Vadodara, Anand and Kheda.
data[data['District_name'].isin({'Vadodara', 'Anand','Kheda'})]
```

```python
#Q. 4) Calculate statewise total population.(Sum Pivot single Columns)
data.groupby('State_name').Population.sum().sort_values(ascending=True)
```

```python
#Q. 5) Calculate population state wise and religion wise. (Sum Pivot Multi Column)
data.groupby('State_name')['Hindus', 'Muslims', 'Christians', 'Sikhs', 'Buddhists', 'Jains'].sum().sort_values(by='Hindus')
```

```python
#Q. 6) How many male workers are there in Maharashtra?(Pivot - Sum - Filter)
data[data.State_name =='MAHARASHTRA']['Male_Workers'].sum()
```

```python
#Q. 7) How to set a column as index?
data = data.set_index('District_code')
data.head(3)
```

```python
#Q. 8) Set a prefix on column name.
data = data.add_prefix('left_')
data.head(3)
```

```python
#Q. 9) Set a suffix on column name.
data = data.add_suffix('_right')
data.head(3)
```
