# Model

### Splitting the Dataset into and Training, Test set and Applying Standard Scalar

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X = sc_X.fit_transform(X)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
sns.distplot(X)
```

```python
sns.distplot(y)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(X_train, y_train)
print("Traiing Score of SVM is: {}".format(log.score(X_train, y_train) * 100))
y_predlog = log.predict(X_test)
am_log = accuracy_score(y_test, y_predlog) * 100
print("Accuracy of Logistic Regression Classifier is: {}%".format(am_log))
print("Confusion Matrix of Logistic Regression Classifier is: \n{}".format(confusion_matrix(y_test, y_predlog)))
print("{}".format(classification_report(y_test, y_predlog)))
```

### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train, y_train)
print("Traiing Score of SVM is: {}".format(nb.score(X_train, y_train) * 100))
y_prednb = nb.predict(X_test)
am_nb = accuracy_score(y_test, y_prednb) * 100
print("Accuracy of Naive Bayes Classifier is: {}%".format(am_nb))
print("Confusion Matrix of Naive Bayes Classifier is: \n{}".format(confusion_matrix(y_test, y_prednb)))
print("{}".format(classification_report(y_test, y_prednb)))
```

### SVM

```python
from sklearn.svm import SVC
svc = SVC(kernel = 'rbf')
svc.fit(X_train, y_train)
print("Traiing Score of SVM is: {}".format(svc.score(X_train, y_train) * 100))
y_predsvm = svc.predict(X_test)
am_svm = accuracy_score(y_test, y_predsvm) * 100
print("Accuracy of SVM Classifier is: {}%".format(am_svm))
print("Confusion Matrix of SVM Classifier is: \n{}".format(confusion_matrix(y_test, y_predsvm)))
print("{}".format(classification_report(y_test, y_predsvm)))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'entropy')
dt.fit(X_train, y_train)
print("Traiing Score of Decision Tree is: {}".format(dt.score(X_train, y_train) * 100))
y_preddt = dt.predict(X_test)
am_dt = accuracy_score(y_test, y_preddt) * 100
print("Accuracy of SVM Classifier is: {}%".format(am_dt))
print("Confusion Matrix of SVM Classifier is: \n{}".format(confusion_matrix(y_test, y_preddt)))
print("{}".format(classification_report(y_test, y_preddt)))
```

### Random Forest

#### Finding the optimal number of estimators

```python
from sklearn.ensemble import RandomForestRegressor
train_accuracy = []
test_accuracy = []

for i in range(100, 600, 100):
    c = RandomForestRegressor(n_estimators = i)
    c.fit(X_train, y_train)
    train_accuracy.append(c.score(X_train, y_train))
    test_accuracy.append(c.score(X_test, y_test))
    
frame_rf = pd.DataFrame({"n_neighbors": range(100, 600, 100), "train_accuracy": train_accuracy, "test_accuracy": test_accuracy})
frame_rf
```

```python
plt.figure(figsize = (10, 7))
plt.plot(range(100, 600, 100), frame_rf["test_accuracy"], label = 'Test')
plt.plot(range(100, 600, 100), frame_rf["train_accuracy"], label = 'Training')
plt.title('Optimal Number of Estimators')
plt.xlabel('Estimators')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 300)
rf.fit(X_train, y_train)
print("Traiing Score of Random Forest is: {}".format(rf.score(X_train, y_train) * 100))
y_predrf = rf.predict(X_test)
am_rf = accuracy_score(y_test, y_predrf) * 100
print("Accuracy of SVM Classifier is: {}%".format(am_rf))
print("Confusion Matrix of SVM Classifier is: \n{}".format(confusion_matrix(y_test, y_predrf)))
print("{}".format(classification_report(y_test, y_predrf)))
```

# Comapring various Models (based on Accuracy)

```python
classifiers = ['Logistic Regression', 'Naive bayes', 'SVM', 'Decision Tree', 'Random Forest']
lst_acc = [am_log, am_nb, am_svm, am_dt, am_rf]
df = pd.DataFrame({'Model': classifiers, 'Accuracy': lst_acc})
px.histogram(data_frame = df, x = 'Model', y = 'Accuracy')
```

```python
kfold = KFold(n_splits=10)
xyz = []
accuracy=[]
classifiers=['SVM','Logistic Regression','Decision Tree', 'Random forest', 'Naive Bayes']
models=[SVC(kernel='rbf'),LogisticRegression(),
        DecisionTreeClassifier(), 
        RandomForestClassifier(n_estimators=300,random_state=0), GaussianNB()]

for i in models:
    model = i
    cv_result = cross_val_score(model,X_train, y_train, cv = kfold, scoring = "accuracy")
    cv_result=cv_result
    xyz.append(cv_result.mean())
    accuracy.append(cv_result)

cv_models_dataframe=pd.DataFrame(xyz, index=classifiers)   
cv_models_dataframe.columns=['CV Mean']    
cv_models_dataframe
cv_models_dataframe.sort_values(['CV Mean'], ascending=[0])
```

```python
box=pd.DataFrame(accuracy,index=[classifiers])
boxT = box.T
plt.figure(figsize = (10, 7))
ax = sns.boxplot(data=boxT, orient="h", palette="Set2", width=.6)
ax.set_yticklabels(classifiers)
ax.set_title('Cross validation accuracies with different classifiers')
ax.set_xlabel('Accuracy')
plt.show()
```
