```python
# Following Notebook consists of SVC algorithm which uses the kernel means types of accuracies to be used for comparing.
# kernels used here are 'rbf' and 'poly'.
# starting with the columns from pregnancies to age using rbf kernel:

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

df=pd.read_csv("../input/diabetes-prediction/Diabetespred.csv")
print(df)

x=df.loc[:,'Pregnancies':'Age']
y=df['Outcome']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)

pqr=SVC(kernel='rbf',random_state=0)
pqr.fit(x_train,y_train)
print(pqr.score(x_train,y_train))
pred=pqr.predict(x_test)
print(y_test)
print(pred)
```
