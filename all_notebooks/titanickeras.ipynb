Ho provato a implementare una rete neurale per il Titanic Dataset.

Con 1000 epoche, dà una precisione superiore all'80%: più di quella che avevo ottenuto con un semplice decision tree.

Il Titanic dataset è un esempio lampante di come il bias nella scelta delle features sia determinante: occorre anche conoscere il campo di applicazione dei modelli.

Sono curioso di capire come era strutturata la soluzione che vinse la corrispondente competizione su Kaggle.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from keras.layers import Dense
from tensorflow.keras.layers import Dense,Dropout
import pandas as pd
import numpy as np
```

```python
model = keras.Sequential()
model.add(Dense(10,input_dim=7,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(15,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(12,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(8,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(5,activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1,activation='sigmoid'))
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
titanic_train = pd.read_csv("../input/titanic-cleaned-data/train_clean.csv")
titanic_test_input = pd.read_csv("../input/titanic-cleaned-data/test_clean.csv")
titanic_test_output = pd.read_csv("../input/titaniccsv/titanic.csv")
```

```python
titanic_train.loc[titanic_train['Sex'] == 'male', 'Sex'] = 0
titanic_train.loc[titanic_train['Sex'] == 'female', 'Sex'] = 1

titanic_test_input.loc[titanic_test_input['Sex'] == 'male', 'Sex'] = 0
titanic_test_input.loc[titanic_test_input['Sex'] == 'female', 'Sex'] = 1

titanic_array_train = titanic_train.values
titanic_test_input_array = titanic_test_input.values
```

```python
titanic_test_output_array = titanic_test_output.values
```

```python
#I've decided to use these features:
#Age (0), Fare(3), Parch(5), Pclass(7), Sex(8), SibSp(9) and family_size(13)

titanic_train_X = titanic_array_train[:,[0,3,5,7,8,9,13]]
titanic_train_X= titanic_train_X.astype('float')
titanic_train_y = titanic_array_train[:,[10]]
titanic_train_y= titanic_train_y.astype('float')

titanic_test_X = titanic_test_input_array[:,[0,3,5,7,8,9,13]]
titanic_test_X = titanic_test_X.astype('float')
titanic_test_y = titanic_test_output_array[:,[1]]
titanic_test_y = titanic_test_y.astype('float')
```

```python
len(titanic_train_y)
```

```python
len(titanic_test_y)
```

```python
model.fit(titanic_train_X, titanic_train_y, batch_size=5, epochs=1000)
```

```python
titanic_prediction = model.predict_classes(titanic_test_X)
```

```python
mismatching_count = 0

#mismatching_count
for i in range(len(titanic_test_y)):
    if titanic_test_y[i] != titanic_prediction[i]:
       mismatching_count+=1
matching = len(titanic_test_y) - mismatching_count
precision = matching/len(titanic_test_y)

print(precision)
```
