```python
ds = pd.get_dummies(df1, columns = ['sex', 'cp', 'fbs', 'restecg', 'exng', 'slp', 'caa', 'thall'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['age', 'trtbps', 'chol', 'thalachh', 'oldpeak']
ds[columns_to_scale] = standardScaler.fit_transform(ds[columns_to_scale])
```

```python
ds.head()
```

```python
y =ds.output.values
X=ds.drop(["output"],axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size = 0.25,random_state = 5)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred = classifier.predict(x_test)
accuracy_score(y_pred,y_test)
```
