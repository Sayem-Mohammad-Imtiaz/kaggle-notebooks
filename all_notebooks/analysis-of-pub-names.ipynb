The top names are interesting - the red lion is a symbol of Scotland but while it is number one in GB it is not in the top 10 for Scotland. The reason probably dates back to King James of Scotland who became king James of England and Ireland after the union of the English and Scottish crowns. People in England wanted to display their loalty to their new Scottish king so many pubs took a symbol of Scotland as their name. While the top three names in Scotland show their loyalty to Britain rather than Scotland.
