Class is a dependable variable . As we can see that all other variable are independent and show that whether that class is 0 or 1 (real or fake).

```python
from sklearn import linear_model, preprocessing
cols = df.columns[0:4]
X=np.array(df[cols].values)
Xs=preprocessing.scale(X)
y = np.array(df['class'])
```

```python
X
```

```python
Xs  #scaled X 
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression
lg=linear_model.LogisticRegression(C=1e5)
lg.fit(Xs,y)
```

```python
y_pre=lg.predict(Xs)
```

```python
acc= np.mean(y_pre==y)
print("Accuracy of the train data :{0:f}".format(acc))
```

Now , we need to check which optimal weight will descriminate fake and genuine note properly .

```python
w=lg.coef_
print(w)
```

```python
W=w.ravel()  #flatten 2 D array to 1 D array 
W
```

```python
from sklearn.model_selection import KFold 
from sklearn.metrics import precision_recall_fscore_support
from sklearn.model_selection import train_test_split
#using K=10 folds for Kcross validation
nfold=10
kf = KFold(n_splits=nfold,shuffle=True)
prec=[]
rec=[]
f1=[]
err_rate=[]
for Itr, Its in kf.split(Xs):
    Xtr=Xs[Itr,:]
    ytr=y[Itr]
    Xts=Xs[Its,:]
    yts=y[Its]
    lg.fit(Xtr,ytr)
    y_pred = lg.predict(Xts)
    precission,recall,f1,_=precision_recall_fscore_support(yts,y_pred,average='binary')
    prec.append(precission)
    rec.append(recall)
    err_rate.append(np.mean(yts != y_pred))
prec=np.mean(prec)
rec=np.mean(rec)
f1=np.mean(f1)
err_mean=np.mean(err_rate)

print('Precision = {0:0.4f}'.format(prec))
print('Recall = {0:0.4f}'.format(rec))
print('F1 = {0:0.4f}'.format(f1))
print('Error Rate = {0:0.4f}'.format(err_mean))
```
