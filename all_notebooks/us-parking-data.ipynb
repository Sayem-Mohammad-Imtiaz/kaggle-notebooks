### How is the overall situation of total parking spaces by (a)Owned/Leased (b) Property type (c) (Within each property type how is owned and leased) (d) By Building Status (e) By Building State?

```python
df.groupby('Owned/Leased')['Total Parking Spaces'].sum()
```

```python
df.groupby('Property Type')['Total Parking Spaces'].sum()
```

```python
df.groupby(['Property Type','Owned/Leased'])['Total Parking Spaces'].sum()
```

```python
df.groupby('Bldg Status')['Total Parking Spaces'].sum()
```

```python
df.pivot_table(columns='Bldg State', values = 'Total Parking Spaces', aggfunc='sum').\
sort_values(by=['Total Parking Spaces'],axis=1, ascending=False).T
```

### In Which Building state parking situation is in excess-put it in order?

```python
df.query('`Bldg Status` == "EXCESS"')[['Bldg State','Total Parking Spaces']].\
    pivot_table(columns='Bldg State', values = 'Total Parking Spaces', aggfunc='count').\
    sort_values(by=['Total Parking Spaces'],axis=1, ascending=False).T
```

### Building cities that have been decommissioned along with total parking.

```python
df.query('`Bldg Status` == "DECOMMISSIONED"')[['Bldg City','Total Parking Spaces']].\
sort_values(['Total Parking Spaces'],ascending=False)
```

### In Which type of property parking space is in excess?

```python
df.query('`Bldg Status` == "EXCESS"')['Property Type'].value_counts()
```

### Where is more active parking i.e either in owned or leased?

```python
df.query('`Bldg Status` == "ACTIVE"')['Owned/Leased'].value_counts()
```

### States With more leased parking similarly with more owned parking.

### Which property type has more active parking?

```python
df.groupby(['Bldg Status','Property Type'])['Total Parking Spaces'].count()
```

### How's the distribution of parking in the city by property type?

```python
sns.histplot(df['Property Type'])
plt.title('Distrubution of Property Types');
```

### How's the distribution of parking in the city by building status and building state?

```python
plt.figure(figsize=(15,8))
sns.countplot(x='Bldg State', hue='Bldg Status', data=df, order=df['Bldg State'].value_counts().index)
plt.suptitle("Distribution of parking in the city by building status and building state")
plt.legend(loc='upper right');
```
