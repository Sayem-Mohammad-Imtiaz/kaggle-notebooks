```python
x=np.array([data['Fuel Consumption Comb (L/100 km)']]).reshape(-1,1)
y=np.array([data['CO2 Emissions(g/km)']]).reshape(-1,1)
```

```python
x,y
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
#print(lm.coef_,lm.intercept_)
lm.fit(x,y)
```

```python
lm.coef_,lm.intercept_
```

**Finding Predicted Values**

```python
y_pred=lm.predict(x)
y_pred
```

```python
np.sum(np.abs(y-y_pred))/x.shape[0]
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
mean_absolute_error(y,y_pred),np.sqrt(mean_squared_error(y,y_pred))
```

**ploting Graph betwwen actual x and y and actual x and predicted y**

```python
plt.scatter(x,y)
plt.scatter(x,y_pred)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2)
x_train.shape,y_train.shape, x_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x_train,y_train)
```

**Mean Absolute error for Train set**

```python
y_train_pred=lm.predict(x_train)
mean_absolute_error(y_train,y_train_pred)
```

**Mean Aboslute errror for Test set**

```python
y_test_pred=lm.predict(x_test)
mean_absolute_error(y_test,y_test_pred)
```
