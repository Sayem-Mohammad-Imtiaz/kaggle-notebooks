> I cant use inch instead of meters OR again pounds instead of kilogram. Therefore i will conversion values in dataframe.

```python
# Pounds to -----> KG

WTList = list(nbadata["WT"])
KGList = [round(float(i)*0.45,1) for i in WTList]
#KGList
nbadata["KGValues"] = KGList
nbadata.head()

# I could use apply() method but i like in this way.
```

```python
# Feet and Inch to Cm

HTData = list(nbadata["HT"])
CmData = [round((float(HTData[i][0])*12 + float(HTData[i][2]))*2.54 , 1) for i in range(0,len(HTData))]
#CmData
nbadata["CmValues"] = CmData
nbadata.head()
```

```python
# Find ratio of cmValues more then 200 to less then 200 in percent

MoreThen200 = len(nbadata[nbadata.CmValues > 200])
LessThen200 = len(nbadata[nbadata.CmValues < 200])

#MoreThen200
#LessThen200 # we can look results if we want

Ratio = round((MoreThen200*100)/LessThen200,3)
Ratio

# we find as a result %68
```

### Bar Plot

```python
data = pd.DataFrame({"team_name_in2016allstar":teamdata,"count_of_team":Teamsin2016})

# visualization
plt.figure(figsize=(12,8))
sns.barplot(x=data.team_name_in2016allstar,y=data.count_of_team)
plt.xticks(rotation=90)
plt.xlabel("Team Names")
plt.ylabel("Count of Teams")
plt.title("Team Name in 2016 All Star")

plt.show()
```

### Pie Plot

```python
labels = teamdata
explode = np.zeros(18,dtype=int)
sizes = Teamsin2016

plt.figure(figsize=(7,7))
plt.pie(sizes,explode=explode,labels=labels,autopct='%1.1f%%')
plt.title('Number of Players Ratio in 2016 All Star',color = 'blue',fontsize = 15)
plt.savefig('graph.png')
plt.show()
```

### Point Plot

> Actually not more fit Point Plot for this data. But I will for repeat and more learn

```python
# Number of Players between of 2010 and 2016 

year_list0 = list(nbadata.Year.unique())
year_list = year_list0[:7]
year_list = year_list[::-1]


hawks = []
lakers = []
bulls = []
clippers = []
clevland = []

for i in year_list:
    x = nbadata[nbadata.Year == i]
    hawks.append(len(x[x.Team == "Atlanta Hawks"]))
    lakers.append(len(x[x.Team == "Los Angeles Lakers"]))
    bulls.append(len(x[x.Team == "Chicago Bulls"]))
    clippers.append(len(x[x.Team == "Los Angeles Clippers"]))
    clevland.append(len(x[x.Team == "Cleveland Cavaliers"]))

df = pd.DataFrame({"Atlanta Hawks":hawks,"Los Angeles Lakers":lakers,"Chicago Bulls":bulls,"Los Angeles Clippers":clippers,"Cleveland Cavaliers":clevland,"Year":year_list})

# Visualization

f,ax1 = plt.subplots(figsize = (16,10))
sns.pointplot(x="Year",y="Atlanta Hawks",data=df, color="orange",alpha=0.9)
sns.pointplot(x="Year",y="Los Angeles Lakers",data=df, color="yellow",alpha=0.9)
sns.pointplot(x="Year",y="Chicago Bulls",data=df, color="blue",alpha=0.4)
sns.pointplot(x="Year",y="Los Angeles Clippers",data=df, color="grey",alpha=0.8)
sns.pointplot(x="Year",y="Cleveland Cavaliers",data=df, color="purple",alpha=0.9)

plt.text(0,4,'Atlanta Hawks',color='orange',fontsize = 17,style = 'italic')
plt.text(0,3.8,'Los Angeles Lakers',color='yellow',fontsize = 17,style = 'italic')
plt.text(0,3.6,'Chicago Bulls',color='blue',fontsize = 17,style = 'italic')
plt.text(0,3.4,'Los Angeles Clippers',color='grey',fontsize = 17,style = 'italic')
plt.text(0,3.2,'Cleveland Cavaliers',color='purple',fontsize = 17,style = 'italic')

plt.xlabel("Years")
plt.ylabel("Players Count in Team")
plt.title("Players Count in Team Between of 2010 and 2016",fontsize=20)

plt.grid()
plt.show()
```

<a id="7"></a>

```python
sns.jointplot("Year","Atlanta Hawks",data=df,color="orange",ratio=2)
sns.jointplot("Year","Los Angeles Lakers",data=df,color="red",ratio=2)
plt.show()
```

### Scatter Plot

```python
df.columns = ["Atlanta_Hawks","Los_Angeles_Lakers","Chicago_Bulls","Los_Angeles_Clippers","Clevland_Cavaliers","Year"]
df.head()
```

```python
# Again we will visualziation for same question.
# ---> Number of player in the AllStar between of 2010 and 2016.

# We can use previous data.


# We should different trace for every different columns
trace1 = go.Scatter(
    x = df.Year,
    y = df.Atlanta_Hawks,
    mode = "markers",
    name = "Atlanta_Hawks",
    marker = dict(color="rgba(120,240,15,0.8)")
)
trace2 = go.Scatter(
    x = df.Year,
    y = df.Los_Angeles_Lakers,
    mode = "markers",
    name = "Los_Angeles_Lakers",
    marker = dict(color="rgba(20,30,240,0.7)")
)
trace3 = go.Scatter(
    x = df.Year,
    y = df.Chicago_Bulls,
    mode = "markers",
    name = "Chicago_Bulls",
    marker = dict(color="rgba(240,1,240,0.9)")
)

data = [trace1,trace2,trace3]

layout = dict(title = 'Number of player in the AllStar between of 2010 and 2016.',
              xaxis= dict(title= 'Years',ticklen= 10,zeroline= False),
              yaxis= dict(title= 'Number of Players',ticklen= 10,zeroline= False)
             )

fig = dict(data=data,layout=layout)
iplot(fig)
```

<a id="10"></a>
### Bar Plot

```python
# Number of Players playing on NBA AllStar between of 2010 and 2016 (for 3 teams)

PlayersTotal = sum(df["Atlanta_Hawks"]),sum(df["Los_Angeles_Lakers"]),sum(df["Chicago_Bulls"])
PlayersTotalList = list(PlayersTotal)

TeamNames = [col for col in df.columns]
TeamNames = TeamNames[:3]

Newdf = pd.DataFrame({"TeamName":TeamNames,"NumberofPlayers":PlayersTotalList})

# visualization
trace = go.Bar(
    x = Newdf.TeamName,
    y = Newdf.NumberofPlayers,
    name = "Number_Of_Player",
    marker = dict(color="rgba(170,130,40,0.8)"),
    text=Newdf.TeamName
)

data=[trace]
layout = dict(title="Number of Players playing on NBA AllStar between of 2010 and 2016",barmode="group",yaxis=dict(title="Number of Player"))
fig = go.Figure(data=data,layout=layout)

iplot(fig)
```

### Pie Plot

```python
# Number of Players playing on NBA AllStar between of 2010 and 2016 (for 3 teams)
PlayersTotal = sum(df["Atlanta_Hawks"]),sum(df["Los_Angeles_Lakers"]),sum(df["Chicago_Bulls"])
PlayersTotalList = list(PlayersTotal)

TeamNames = [col for col in df.columns]
TeamNames = TeamNames[:3]

Newdf = pd.DataFrame({"TeamName":TeamNames,"NumberofPlayers":PlayersTotalList})

fig = {
    "data": [
        {
            "values":Newdf.NumberofPlayers,
            "labels":Newdf.TeamName,
            "domain":{"x":[0,.5]}, # locaiton of x plane
            "name":"Number of Players Rate",
            "hoverinfo":"label+percent+name", # what we wanna to see when mouse hoover of pie
            "hole": .2,
            "type": "pie" # type of plot
        },
    ],
    
    "layout":{
        "title":"Number of Players playing on NBA AllStar between of 2010 and 2016",
        "annotations":[ 
            {
                "font":{"size":18},
                "showarrow":False,
                "text": "Number of Player",
                "x": 0.40,
                "y": 1
            }
        ]
    }
}

iplot(fig)
```
