So, yes let's come to the main part! yes prediction, we will build 5 differnt regression  model and let's chech which perform best(Thanks to scikit learn which made coding easy)  

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import StratifiedKFold
```

Let's split our data into training and test with ratio 7:3 and evaluate our model 

```python
from sklearn.model_selection import train_test_split
X=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
def model_score(model,X_train,X_test,y_train,y_test):
    model.fit(X_train,y_train)
    return model.score(X_test,y_test)
```

```python
lr=model_score(LogisticRegression(),X_train,X_test,y_train,y_test)
svc=model_score(SVC(),X_train,X_test,y_train,y_test)
knn=model_score(KNeighborsClassifier(),X_train,X_test,y_train,y_test)
nb=model_score(GaussianNB(),X_train,X_test,y_train,y_test)
dt=model_score(DecisionTreeClassifier(),X_train,X_test,y_train,y_test)
```

```python
print("logitic score:",lr)
print("svc score:",svc)
print("knn score:",knn)
print("naive score:",nb)
print("decision tree score:",dt)
```

Looks like for our dataset Logestic regression performed well.
