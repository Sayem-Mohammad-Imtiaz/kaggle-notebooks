```python
cars_data = pd.read_csv("/kaggle/input/simple-peugeot207-pricelist/Peugeot207.csv")
cars_data
```

```python
cars_data["color"].unique()
```

```python
colors = pd.get_dummies(cars_data["color"])
cars_data = pd.concat([cars_data,colors],axis=1)
print(colors)
cars_data
```

```python
gearbox_types = pd.get_dummies(cars_data["gearbox"])
print(gearbox_types)
cars_data = pd.concat([cars_data,gearbox_types],axis=1)
cars_data
```

```python
options = pd.get_dummies(cars_data["option"])
print(options)
cars_data = pd.concat([cars_data,options],axis=1)
cars_data
```
