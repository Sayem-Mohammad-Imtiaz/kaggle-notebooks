## Vendas por ano

```python
vendas_por_ano = df.groupby(df.index.year)['total_item'].sum()
vendas_por_ano
```

```python
layout = {'title': 'Vendas por ano', 'xaxis': {'title': 'Ano'}, 'yaxis': {'title': 'Vendas($)'}}
data = [go.Scatter(x=vendas_por_ano.index, y=vendas_por_ano, 
                   mode='lines+markers',
                   marker={'color': 'red'})]
fig = {'data': data, 'layout': layout}
iplot(fig)
```

## Vendas por dia da semana

```python
vendas_por_dia = df.groupby(df.index.weekday)['total_item'].sum()
vendas_por_dia
```

```python
dias = {0: 'Segunda', 1: 'Terça', 2: 'Quarta', 3: 'Quinta', 4: 'Sexta', 5: 'Sábado', 6: 'Domingo'}
```

```python
receita_por_dia = df.groupby([df.index.weekday]).price_x.sum()
layout = dict(title = 'Vendas por dia da semana',
              xaxis = dict(title = 'Dia da semana'),
              yaxis = dict(title = 'Vendas ($)'),
              )
data = [go.Scatter(x=list(map(lambda x: dias[x], receita_por_dia.index)),
                   y=receita_por_dia, mode='lines+markers',
                 marker=dict(color='blue'))]
fig = dict(data=data, layout=layout)
iplot(fig)
```

## Soma das vendas por horas do dia

```python
vendas_por_hora = df.groupby(df.index.hour)['total_item'].sum()
vendas_por_hora
```

```python
layout = dict(title = 'Soma das vendas por hora do dia',
              xaxis = dict(title = 'Horas do dia'),
              yaxis = dict(title = 'Vendas ($)'),
              )
data = [go.Scatter(x=vendas_por_hora.index,
                   y=vendas_por_hora, mode='lines+markers',
                 marker=dict(color='blue'))]
fig = dict(data=data, layout=layout)
iplot(fig)
```

```python
count_vendas_por_hora = df.groupby(df.index.hour)['total_item'].count()
count_vendas_por_hora
```

```python
layout = dict(title = 'Contagem de vendas por hora do dia',
              xaxis = dict(title = 'Horas do dia'),
              yaxis = dict(title = 'Contagem (#)'),
              )
data = [go.Scatter(x=count_vendas_por_hora.index,
                   y=count_vendas_por_hora, mode='lines+markers',
                 marker=dict(color='blue'))]
fig = dict(data=data, layout=layout)
iplot(fig)
```
