# **Getting details for Sydney**

Since I am interested in rain prediction for Sydney-Australia, it will be better for me to select the details for Sydney. Also I selected the weather parameters at 3pm because that is when the weather for tomorrow may likely be determined by human observation.








```python
dataset2=dataset.loc[(dataset['Location']=='Sydney'),['Temp3pm','WindSpeed3pm','Humidity3pm','Pressure3pm','RainTomorrow']]
    
```

```python
#dispalying the dataset

dataset2.head()
```

```python
#details about the dataset

dataset2.info()
```

```python
#dealing with null values

dataset3=dataset2.dropna()
```

```python
dataset3.info()
```

```python
#splitting the dataset into x and y

X = dataset3.iloc[:, :-1].values
y = dataset3.iloc[:, -1].values
```

```python
X
```

```python
y
```

```python
#encoding the dependent variable

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
#displaying the encoded y

y
```

# **Splitting the dataset into the Training set and Test set**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
print(X_train)
```

```python
print(y_train)
```

```python
print(X_test)
```

```python
print(y_test)
```

# **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
print(X_test)
```

# **Training the SVM model on the Training set**

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(X_train, y_train)
```

# **Predicting the Test set results**

```python
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# **Making the Confusion Matrix**

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
