```python
# This data describes temperature and humidity changes for each of the two layers (A and B) in different cubes
# Below I imported the data, looked at it, deleted unnecessary, changed the column names and data types
# One of the goals of this exercise is to find cubes that behave differently from most
# Next step will be visualizing temperature and humidity changes for each cube

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv('/kaggle/input/work-for-parmavir/cubes.csv', sep=',')
data
```

```python
data.describe()
```

```python
data.info()
```

```python
df = data.drop(['Unnamed: 0'],axis=1)
df
```

```python
df.rename({'Temperature Layer A': 'Temperature_Layer_A'}, axis=1, inplace=True)
df.rename({'Temperature Layer B': 'Temperature_Layer_B'}, axis=1, inplace=True)
df.rename({'Humidity Layer A': 'Humidity_Layer_A'}, axis=1, inplace=True)
df.rename({'Humidity Layer B': 'Humidity_Layer_B'}, axis=1, inplace=True)
df.head()
```

```python
df['Temperature_Layer_A'] = df.Temperature_Layer_A.str.replace('°C', '').astype(float).round(1)
df['Temperature_Layer_B'] = df.Temperature_Layer_B.str.replace('°C', '').astype(float).round(1)
df['Humidity_Layer_A'] = df.Humidity_Layer_A.str.replace('g/m3', '').astype(float).round(5)
df['Humidity_Layer_B'] = df.Humidity_Layer_B.str.replace('g/m3', '').astype(float).round(5)
df.head()
```

```python
df['Timestamp'] = df['Timestamp'] - df['Timestamp'].min()
df.Timestamp.describe()
```
