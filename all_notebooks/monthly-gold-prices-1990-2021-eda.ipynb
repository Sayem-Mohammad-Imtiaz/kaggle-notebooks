# Part 1
**General wisualizations**

Firstly, let's show price of goold changed by time.

```python
def line_plot(data : pd.DataFrame, column : str):
    plt.figure(figsize=(5,4))
    plt.plot(data["Date"], data[column])
    plt.title(column)
    plt.show()
```

```python
for column in list(data.columns)[1:]:
    line_plot(data, column)
```

Now let see see factual numerical difference betwee first and last measurements of gold in every currency.

```python
currency_cols = list(data.columns)[1:]
dataset_len = len(data)
factual_increases = [ data[col][dataset_len-1]-data[col][0] for col in currency_cols ]
factual_increases_data_package = pd.DataFrame(factual_increases, index=currency_cols)

plt.figure(figsize=(8, 12))
#plt.plot(np.log(factual_increases), currency_cols, "pb", xlabel="logs of factual increases", ylabel="currency")
plt.plot(np.log(factual_increases), currency_cols, ">b")
plt.xlabel("logs of factual increases")
plt.ylabel("currency")
plt.show()
```
