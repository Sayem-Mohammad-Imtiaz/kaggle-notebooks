# 5) Cleaning Data

```python
#mapping data those we need
map1 = {"high school": 1, "some high school": 1,
        "associate's degree": 2,
        "some college": 3,
        "bachelor's degree": 4,
        "master's degree": 5}
data['parental level of education']  = data['parental level of education'].map(map1)

map2 = {"free/reduced": 0,
        "standard": 1}
data['lunch']  = data['lunch'].map(map2)

map3 = {"none": 0,
        "completed": 1}
data['test preparation course']  = data['test preparation course'].map(map3)

map4 = {"female": 0,
        "male": 1}
data['gender']  = data['gender'].map(map4)

map5 = {"group A": 1,
        "group B": 2,
        "group C": 3,
        "group D": 4,
        "group E": 5}
data['race/ethnicity']  = data['race/ethnicity'].map(map5)
```

```python
data.head()
```

```python
# Dropping unnecessary features
dataDrop = data.drop(['math score','reading score','writing score', 'math_status', 'reading_status','writing_status','total','average'], axis=1)
dataDrop.head()
```
