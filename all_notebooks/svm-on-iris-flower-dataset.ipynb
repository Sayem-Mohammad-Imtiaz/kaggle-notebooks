# Build an SVM model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
X = iri.drop('species', axis = 1)
y = iri['species']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 101)
```

```python
from sklearn.svm import SVC
```

```python
mod_1 = SVC()
```

```python
mod_1.fit(X_train, y_train)
```

```python
pred_1 = mod_1.predict(X_test)
```

```python
print(classification_report(y_test, pred_1))
```

Running a grid search, an advantageous feature of sklean to select appropriate 'C' and 'gamma' values

```python
from sklearn.model_selection import GridSearchCV
```

```python
para_grid = {'C':[0.1, 1, 10, 100], 'gamma':[1, 0.1, 0.01, 0.001]}
```

```python
grid = GridSearchCV(SVC(), para_grid, verbose=3)
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid_predictions = grid.predict(X_test)
```

```python
print(classification_report(y_test, grid_predictions))
```

Well it looks like this didn't make quite a change in the result.
But this grid search be used for SVM on further huge datasets to get accurate and beter results.
