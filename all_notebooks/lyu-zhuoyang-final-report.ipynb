# Create a global map for COVID Vaccinations (Ratio (in ppm) between vaccination number and total population for the current date)

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
print(df.columns)

fig = px.choropleth(df,locations='country',locationmode='country names',color='daily_vaccinations_per_million',hover_name='country',animation_frame='date',color_continuous_scale='OrRd',range_color=(0.0,15000.0))
fig.update_layout(title_text='Daily ratio(ppm) between vaccination number and the country total population')
fig.show()

plotly.offline.plot(fig,filename='vaccinations.html')
```

I will describe what I am visualizing and explain the reason why I selected it.
First, I plot the animation of Daily ratio(ppm) between vaccination number and the country total population. So it shows that how many vaccinations per million for one day, and it also shows the change of the daily ratio.

The reason why I choose it is that I would like to see that would the country provide enough vaccine, and would citizen prefer to vaccinate, and the change of citizen consideration. I think ratio (in ppm) between vaccination number and total population for the current date can show what I want to see. If the country cannot provide enough vaccine, the ratio of course not to be so high. And if citizen do not like/afraid to vaccinate, the ratio would not be high. What's more, it is a daily data, we can see the change of the ratio which can show us, how citizen minds change and how hard the government try to get or product vaccine.

# Create a global map of earthquake occurrences worldwide.
