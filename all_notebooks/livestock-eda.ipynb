Importing all the necessary Python libraries.

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import matplotlib.colors as color
import seaborn as sns
%matplotlib inline
```

Loading the data. 

```python
df = pd.read_csv("../input/20th-livestock-census-india/Livestock _census_20.csv")
```

```python
#creating a dataframe with values grouped according to states
state = df.groupby("state_name").sum().reset_index()
state.drop(["ref_district_id", "ref_state_id"], axis = 1 , inplace  = True)

#converting the data in float value
li = state.columns.tolist()
state[li[1:]] = state[li[1:]].astype(float)
state
```

To visualize the popuation of all animals in all states, let us plot a stacked bar chart. This should help us answer questions, such as, Which state have the maximum number of livestock animals? Which state lead in the population of individual animal? etc.

```python
c = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf','#8e0062']
ax = state.plot.barh( x = "state_name", stacked = True, figsize=(20,20), color = c)
ax.set(title = "Livestock Population in each state", ylabel = "States", xlabel = "Population")
plt.savefig('ax.png')
```

As we can see Tamil Nadu has the maximum nummber of livstock animals, out which majority are poultry animals, infact Tamil Nadu has highest number of poultry animals followed by Andhra Pradesh. Also we can see Uttar Pradesh has the highest number of Buffalo, Assam has highest popualtion of pigs. Maximum number of cattle and goats are in West bengal and Rajasthan respectively. You can also find that there are only a few states where poultry population is less than all other animals overall.

Now,lets visualize the population distribution at district level.

```python
#setting the figure size for all further charts
sns.set(rc={'figure.figsize':(20,20)})

```

```python
ax1 = sns.histplot(x = "cattle", data = df, log_scale = True, kde = True)
ax1.set(ylabel = "No. of Districts", xlabel = "Cattle Population", title = "Cattle Population distribution")
plt.axvline(df["cattle"].mean(), color = "purple")
plt.savefig('ax1.png')
```

```python
ax2 = sns.histplot(x = "Total_poultry", data = df, log_scale = True, kde = True)
ax2.set(ylabel = "No. of Districts", xlabel = "Poultry Population", title = "Poultry Population distribution")
plt.axvline(df["Total_poultry"].mean(), color = "purple")
plt.savefig('ax2.png')
```

```python
ax3= sns.histplot(x = "goat", data = df, log_scale = True, kde = True)
ax3.set(ylabel = "No. of Districts", xlabel = "Goat Population", title = "Goat Population distribution")
plt.axvline(df["goat"].mean(), color = "purple")
plt.savefig('ax3.png')
```

Here we trying to see the distribution of animals in India by plotting their population as a histogram.
After plotting the population of Cattle, Poultry animals and Goats, we can say that in majority of districts in India the population of cattle is between 100000 to 900000 and the mean, which is the purple line is around 300000.
The distribution for poultry is more spread out, majority districts having population in the range of 90000 to 1100000 with most districts having a population of 1000000.
Similarly, for goats the range is 90000 to 600000 and the mean is around 200000.


```python
ax4 = sns.boxplot(x = "buffalo",y = "state_name", data = df)
ax4.set(ylabel = "No. of Districts", xlabel = "Buffalo Population", title = "Buffalo Population distribution in each state")
plt.savefig('ax4.png')
```

```python
ax5 = sns.boxplot(x = "donkey",y = "state_name", data = df)
ax5.set(ylabel = "No. of Districts", xlabel = "Donkey Population", title = "Donkey Population distribution in each state")
plt.savefig('ax5.png')
```

In the above boxcharts, we can see the population distribution in each state. In case of buffalo, they are the highest in number in Uttar Pradesh, where majority of districts have around 400000 buffalo population. But the individual district which have the highest number of buffalo is in Gujrat.
Also, by observing the chart for donkeys, we can see that majority of the districts in India have less than 2000 donkey, While the highest number of donkeys in a district is around 5000.
