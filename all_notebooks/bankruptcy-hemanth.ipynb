```python
df= pd.read_csv('/kaggle/input/company-bankruptcy-prediction/data.csv')
df.head()
```

```python
df.describe()
```

```python
df.isna().count()
```

```python
cols= df.columns
num_cols = df._get_numeric_data().columns
char_cols=list(set(cols) - set(num_cols))
char_cols
```

```python
from sklearn.impute import SimpleImputer
si= SimpleImputer()
df.loc[:,:]= si.fit_transform(df.loc[:,:])
df
```

```python
df.corr()
```

```python
X= df.iloc[:,1:]
y= df.iloc[:,0]
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=int(len(cols)/4))
X = pca.fit_transform(X.values)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2)
```

```python
!pip install auto-sklearn
```

```python
import autosklearn.classification
model = autosklearn.classification.AutoSklearnClassifier(time_left_for_this_task=5*60, per_run_time_limit=30, n_jobs=8)
model.fit(X_train, y_train)
```

```python
print(model.sprint_statistics())
```

```python
from sklearn.metrics import accuracy_score
y_hat = model.predict(X_test)
acc = accuracy_score(y_test, y_hat)
print("Accuracy: %.3f" % acc)
```

```python
from xgboost import XGBClassifier
clf=XGBClassifier()
clf.fit(X_train,y_train)
y_pred= clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("Accuracy: %.3f" % acc)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
clf.fit(X_train,y_train)
y_pred= clf.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("Accuracy: %.3f" % acc)
```
