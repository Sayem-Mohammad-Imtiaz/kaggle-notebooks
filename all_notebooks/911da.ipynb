```python
df["Reason"]=df["title"].apply(lambda x: x.split(":")[0])
```

Extracted reason category from title

```python
df.head()
```

```python
df["Reason"].value_counts()
```

let's see how it works

```python
sns.countplot(x="Reason",data=df)
```

Checked out the count of each reason

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
```

Transformed data type from str to timestamp

```python
df['Hour'] = df['timeStamp'].apply(lambda time: time.hour)
df['Month'] = df['timeStamp'].apply(lambda time: time.month)
df['Day of Week'] = df['timeStamp'].apply(lambda time: time.dayofweek)
```

Inserted new columns, separating hour, month, and day of week

```python
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day of Week']=df["Day of Week"].map(dmap)
```

Transformed 'day of week' to integer

```python
sns.countplot(x="Day of Week",data=df,hue="Reason")
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.9))
```

Checked out different reason's appearance each day of week

```python
sns.countplot(x='Month',data=df,hue='Reason',palette='viridis')

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

Checked out Month

```python
df["Date"]=df["timeStamp"].apply(lambda time: time.date())
df.head()
```

Created new col 'Date' from 'timeStamp'

```python
plt.figure(figsize=(16,4))
dd=df.groupby("Date").count()
dd["e"].plot()
plt.tight_layout()
```

Checked out the trend of 911 calls by time series

```python
plt.figure(figsize=(16,4))
Traffic=df[df["Reason"]=="Traffic"].groupby("Date").count()
Traffic["e"].plot().set_title("Traffic")
```

Checked out the trend of 'traffic' 911 calls by time series

```python
plt.figure(figsize=(16,4))
Fire=df[df["Reason"]=="Fire"].groupby("Date").count()
Fire["e"].plot().set_title("Fire")
```

Checked out the trend of 'fire' 911 calls by time series

```python
plt.figure(figsize=(16,4))
EMS=df[df["Reason"]=="EMS"].groupby("Date").count()
EMS["e"].plot().set_title("EMS")
```

Checked out the trend of 'EMS' 911 calls by time series
Clearly found out there's a significant trend than 'fire' and 'traffic'

```python
df.groupby(by=["Day of Week","Hour"]).count()["e"].unstack(level=-1)
```

Checked out every hour's 911 calls in each day of week

```python
newdf=df.groupby(by=["Day of Week","Hour"]).count()["e"].unstack(level=-1)
plt.figure(figsize=(12,6))
sns.heatmap(data=newdf,cmap="viridis")
```

Found out the peak of 911 calls by different hour in each day of week

```python
sns.clustermap(newdf)
```
