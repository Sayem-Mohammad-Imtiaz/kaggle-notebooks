```python
uf=pd.read_csv('../input/user_features.csv').values
```

```python
nm=np.zeros([8874627,14],dtype=int)
for i in range(8874627):
    nm[i][0]=i
```

```python
start=time.time()
l=0
o=0
for i in range(8874626):
  if(uf[o][0]==nm[i][0]):
    for j in range(13):
        nm[i][j+1]=uf[o][j+1]
    o=o+1
stop=time.time()
print(stop-start)    
```

```python

ntr=np.zeros([4503568,27],dtype=int)

```

```python
start=time.time()
for i in range(4503568):
    ind1=df2[i][0]
    ind2=df2[i][1]
    for j in range(13):
        ntr[i][j]=nm[ind1][j+1]
        ntr[i][j+13]=nm[ind2][j+1]
stop=time.time()
print(stop-start)
        
```

```python
for i in range(4503568):
    ntr[i][26]=df2[i][2]
 
```

```python
X=ntr[:,:26]
y=ntr[:,26]
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
classifier = Sequential()
classifier.add(Dense(units = 26, kernel_initializer = 'uniform', activation = 'tanh', input_dim = 26))
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'tanh'))
classifier.add(Dropout(rate=0.5))
classifier.add(Dense(units = 1, kernel_initializer = 'uniform',activation='linear'))
classifier.compile(optimizer = 'adadelta', loss = 'mean_squared_error', metrics = ['accuracy'])
```

```python
classifier.fit(X, y, batch_size = 1000, epochs = 100,validation_split = 10000)
```

```python
tst=pd.read_csv('../input/test.csv').values
otst=np.zeros([11776968,26],dtype=int)
```

```python
for i in range(11776968):
    ind1=tst[i][1]
    ind2=tst[i][2]
    for j in range(13):
        otst[i][j]=nm[ind1][j+1]
        otst[i][j+13]=nm[ind2][j+1]
```

```python
ypred=classifier.predict(otst)
```
