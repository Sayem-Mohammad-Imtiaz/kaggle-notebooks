## Import libraries

```python
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
```

```python
df = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13,df14],ignore_index=True)
df.shape
```
