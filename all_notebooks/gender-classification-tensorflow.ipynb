```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
data=pd.read_csv('/kaggle/input/gender-classification/Transformed Data Set - Sheet1.csv')
data
```

```python
fav_clr,clr_count=np.unique(data['Favorite Color'],return_counts=True)
plt.pie(clr_count , labels = fav_clr,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
fav_msc,msc_count=np.unique(data['Favorite Music Genre'],return_counts=True)
plt.pie(msc_count , labels = fav_msc,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
fav_bvrg,bvrg_count=np.unique(data['Favorite Beverage'],return_counts=True)
plt.pie(bvrg_count , labels = fav_bvrg,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
fav_sft,sft_count=np.unique(data['Favorite Soft Drink'],return_counts=True)
plt.pie(sft_count , labels = fav_sft,autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
def ann():
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(16,activation='relu',input_dim=20),
        tf.keras.layers.Dense(8,activation='relu'),
        tf.keras.layers.Dense(2,activation='softmax')
    ])
    model.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
    return model
```

```python
data_dum=pd.get_dummies(data)
data_dum
```

```python
data_dum=data_dum.to_numpy()
```

```python
X=data_dum[:,:-2]
y=data_dum[:,-2:]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
model=ann()
history=model.fit(X_train,y_train,epochs=32,batch_size=1)
```

```python
history.history.keys()
```

```python
plt.plot(history.history['accuracy'])
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.ylabel('loss')
plt.xlabel('epoch')
plt.show()
```

```python
y_pred=model.predict(X_test)
```

```python
y_test_inverse=np.argmax(y_test,axis=1)
y_pred_inverse=np.argmax(y_pred,axis=1)
```

```python
cm=confusion_matrix(y_test_inverse,y_pred_inverse)
ac=accuracy_score(y_test_inverse,y_pred_inverse)
cm,ac
```
