### Atlas of Rural and Small-Town America

- This notebook shows how to extract a single sub category from the People data and visualize it.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Let's load the People table and take a look

```python
# Load the csv and look at the head
people = pd.read_csv("/kaggle/input/atlas-of-rural-and-smalltown-america/People.csv")
people.head()
```

#### This table has population and migration counts for the US and each state/county.

- Let's take a look at the population of all counties in a single state. We'll choose the state of Tennessee.

```python
# Get the rows only for the state of TN, ignore the first row, which contains the entire state data.
people_in_TN = people[people['State'] == "TN"]
pop_2018 = people_in_TN[['County','TotalPopEst2018']].iloc[1:]

# Sort the population values in descending order.
pop_2018 = pop_2018.sort_values(['TotalPopEst2018'], ascending=False).reset_index(drop=True)
pop_2018.head()
```

```python
# Plot the population by county for TN
plt.figure(figsize = (8,16))
sns.barplot(y="County", x="TotalPopEst2018", data=pop_2018);
```
