Ä°t's very small dataset. We have to use all features without any restriction.

```python
x=data.iloc[:,0:4]
y=data.iloc[:,4:]
```

With this small dataset. I think decision tree is ideal

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20 ,random_state=2002)
```

```python
from sklearn import tree
cls=tree.DecisionTreeClassifier(criterion="entropy")
cls.fit(x_train,y_train)
ypred=cls.predict(x_test)
import sklearn.metrics as metrik
print(metrik.accuracy_score(y_true=y_test,y_pred=ypred))
print(metrik.confusion_matrix(y_true=y_test,y_pred=ypred)
```

```python
print(metrik.classification_report(y_true=y_test,y_pred=ypred))
```

Seems like work  fine but these evaluation is not ideal

```python
from sklearn.tree import export_graphviz
export_graphviz(cls,feature_names=x.columns.values,out_file="tree.dot")
```
