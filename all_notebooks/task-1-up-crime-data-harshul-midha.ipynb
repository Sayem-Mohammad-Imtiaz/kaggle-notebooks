## Crimes against Women

```python
women=np.array(df[11].value_counts())
women2=df[12].value_counts()
x1=women[0]-df[8].value_counts()[0]
x2=466-women[0]-(466-women2[0])
x3=466-women2[0]
explode=(0,0.1,0.1)
labels=["Crimes against women under one section","Crimes against women under two sections","Crimes against women under three sections"]
sizes=[x1,x2,x3]
plt.pie(sizes,explode=explode,autopct='%1.2f%%',shadow=True,startangle=120)
plt.legend(labels,loc="best")
plt.show()
```

It can be seen from the above pie chart that approximately 34% of crimes against women come under multiple sections together.

```python
groupings=['8,10','4,7','1,10','7,10','3,7','5,6','2,3','7,9','2,6','4,8','3,8','7,8','1,4','6,8','8,9','4,7,10','3,7,10','2,3,6','1,4,10']
count=[0 for i in range(len(groupings))]
for i in range(len(groupings)):
    cut=groupings[i].split(',')
    if len(cut)==2:
        for j in range(len(df)):
            if str(df.iloc[j,8])==cut[0] and str(df.iloc[j,11])==cut[1] and df.iloc[j,12]==0:
                count[i]=count[i]+1
    elif len(cut)==3:
        for j in range(len(df)):
            if str(df.iloc[j,8])==cut[0] and str(df.iloc[j,11])==cut[1] and str(df.iloc[j,12])==cut[2]:
                count[i]=count[i]+1
```

```python
fig3=plt.figure(figsize=(16,9))
plt.bar(groupings,count,width=0.4)
plt.xlabel("Groupings possible")
plt.ylabel("Number of instances")
plt.title("Number of instances of each grouping possible")
plt.show()
```

It can be seen from the barplot that crimes often come under section 7,10 together, and also sometimes under section 8,10 together. Crimes also come under section 2,3,6 together quite a few times and also quite a few times under section 3,8 together.
Note that in the above barplot 4,7,10 is not already included in 7,10 i.e. no crime is counted twice.
