```python
import pandas as pd
import numba as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("../input/average-rainfall-and-temperature-in-uk20102019/avg_rainfalltemp in UK - Sheet1.csv")
```

```python
plt.figure(figsize=(12,7))
plt.style.use("fivethirtyeight")
sns.lineplot(x='Year', y='Avg rainfall(in mm)',data=data, hue="Type of period")
```

```python
plt.figure(figsize=(12,7))
plt.style.use("fivethirtyeight")
sns.lineplot(x='Year', y='Avg temp(in centigrade)',data=data, hue="Type of period")
```

```python
plt.figure(figsize=(20,7))
plt.style.use("ggplot")
sns.lineplot(x='Period', y='Avg rainfall(in mm)',data=data, hue="Year")
```
