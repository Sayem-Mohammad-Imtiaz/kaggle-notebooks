**1.4 Проведите необходимую предобработку данных (preprocessing). Для построения моделей с помощью метрических методов все признаки должны быть закодированы числами.**

Категориальных переменных здесь нет, следовательно перекодировка не нужна, однако график распределения target-переменной немного смещен вправо и можно прологорифмировать её.

```python
df['quality_new'] = np.log(df['quality'])
df_target_new = df['quality_new']
print(df['quality_new'])
print(df['quality_new'])
```

Можно воспользоваться классом StandardScaler для масштабирования. Как подготовка датафрейм для дальнейшего разделения на выборки - сохраним target-переменную отдельно.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_scaled = df.drop('quality', axis = 1)
df_scaled = df.drop('quality_new', axis = 1)
df_scaled_res = scaler.fit_transform(df_scaled)
df_scaled_res
```

**1.5 Разбейте набор данных на обучающую и валидационную (тестовую) выборки
с помощью метода train_test_split.**

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(df_scaled_res, 
                                                      df_target_new, 
                                                      test_size=0.3, 
                                                      random_state=50)
```

**1.6 Обучите алгоритм регрессии**

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor(n_neighbors=50)
```

```python
knn.fit(X_train, y_train)
y_pred = knn.predict(X_valid)
```

Оценим качество модели с помощью mean_squared_error

```python
knn.score(X_valid, y_valid)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_valid, y_pred)
```

Модель можно считать качественной тк значение mean_squared_error довольно маленькая, а score достаточно большое.
