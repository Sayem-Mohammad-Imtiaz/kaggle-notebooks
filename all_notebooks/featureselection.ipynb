```python
from sklearn.datasets import fetch_california_housing
data= fetch_california_housing()
data
```

```python
X= data['data']
col_names=data['feature_names']
y=data['target']
y
```

```python
df= pd.DataFrame(X,columns=col_names)
df.head()
```

```python
df.loc[:,'MedInc_Sqrt']=df.MedInc.apply(np.sqrt)
df.corr()
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
model=LinearRegression()
rfe=RFE(estimator=model,n_features_to_select=4)
rfe.fit(X,y)
X_trans=rfe.transform(X)
pd.DataFrame(X_trans)
```

```python
from sklearn.datasets import load_diabetes
from sklearn.ensemble import RandomForestRegressor
data=load_diabetes()
X=data['data']
col_names= data['feature_names']
y=data['target']
model=RandomForestRegressor()
model.fit(X,y)
import matplotlib.pyplot as plt
imp= model.feature_importances_
idxs= np.argsort(imp)
plt.barh(range(len(idxs)),imp[idxs],align='center')
plt.yticks(range(len(idxs)),[col_names[i] for i in idxs])
plt.show()
```

```python
from sklearn.feature_selection import SelectFromModel
sf=SelectFromModel(estimator=model)
X_trans=sf.fit_transform(X,y)
support=sf.get_support()
print([x for x,y in zip(col_names,support) if y==True])
```
