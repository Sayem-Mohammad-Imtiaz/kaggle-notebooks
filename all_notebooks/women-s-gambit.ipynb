#Gambit

A gambit (from ancient Italian gambetto, meaning "to trip") is a chess opening in which a player, more often White, sacrifices material, usually a pawn, with the hope of achieving a resulting advantageous position. Some well-known examples are the King's Gambit, Queen's Gambit, and Evans Gambit. A gambit used by Black may also be called a gambit, e.g. the Latvian Gambit , or Englund Gambit; but is sometimes called a "countergambit", e.g. the Albin Countergambit and Greco Countergambit (an old-fashioned name for the Latvian Gambit).https://en.wikipedia.org/wiki/Gambit

```python
#Code from Gabriel Preda
#plt.style.use('dark_background')
def plot_count(feature, title, df, size=1):
    f, ax = plt.subplots(1,1, figsize=(4*size,4))
    total = float(len(df))
    g = sns.countplot(df[feature], order = df[feature].value_counts().index[:20], palette= ('summer'))
    g.set_title("Number and percentage of {}".format(title))
    if(size > 2):
        plt.xticks(rotation=90, size=8)
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2.,
                height + 3,
                '{:1.2f}%'.format(100*height/total),
                ha="center") 
    plt.show()
```

#The World Chess Federation (FIDE)

The World Chess Federation (FIDE) divides time controls for chess into "classical" time controls, and the fast chess time controls. Games played faster than these time controls (120 minutes per player based on a 60-move game) can be rated for rapid and blitz if they comply with the time controls for those categories.

Players of fast and blitz chess are exempt from the requirement to record their moves onto a scoresheet . The arbiter or their assistant is responsible for the recording in competitions . Electronic recording is preferred. https://en.wikipedia.org/wiki/Fast_chess#:~:text=increments%20per%20move.-,Rapid%20(FIDE)%2C%20quick%20(USCF)%20or%20active,time%20increments%20for%20each%20move.&text=Rapid%20chess%20was%20called%20active%20chess%20by%20FIDE%20between%201987%20and%201989.

```python
#Code by Gabriel Preda https://www.kaggle.com/gpreda/explore-the-rainforest-soundscape
def plot_feature_distribution(df, feature, feature2, title, kde_mode=False, hist_mode=True):
    f, ax = plt.subplots(1,1, figsize=(12,6))
    for item in list(df[feature2].unique()):
        d_df = df.loc[df[feature2]==item]
        try:
            sns.distplot(df[feature], kde=kde_mode, hist=hist_mode, label=item)
        except:
            pass
    plt.legend(labels=list(df[feature2].unique()), bbox_to_anchor=(1, 1), loc='upper right', ncol=2)
    plt.title(title)
    plt.show()
```

```python
# Lets first handle numerical features with nan value
numerical_nan = [feature for feature in df.columns if df[feature].isna().sum()>1 and df[feature].dtypes!='O']
numerical_nan
```

```python
df[numerical_nan].isna().sum()
```

```python
## Replacing the numerical Missing Values

for feature in numerical_nan:
    ## We will replace by using median since there are outliers
    median_value=df[feature].median()
    
    df[feature].fillna(median_value,inplace=True)
    
df[numerical_nan].isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

#fill in mean for floats
for c in df.columns:
    if df[c].dtype=='float16' or  df[c].dtype=='float32' or  df[c].dtype=='float64':
        df[c].fillna(df[c].mean())

#fill in -999 for categoricals
df = df.fillna(-999)
# Label Encoding
for f in df.columns:
    if df[f].dtype=='object': 
        lbl = LabelEncoder()
        lbl.fit(list(df[f].values))
        df[f] = lbl.transform(list(df[f].values))
        
print('Labelling done.')
```
