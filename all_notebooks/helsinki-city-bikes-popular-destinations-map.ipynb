# **EDA Helsinki City Bikes**

```python
import pandas as pd
import matplotlib.pyplot as plt

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('/kaggle/input/helsinki-city-bikes/database.csv')
df.head()
```

```python
df.info()
```

### Getting just coordinates to have map

```python
l_l = df[['departure_latitude','departure_longitude']].drop_duplicates()
ax = l_l.plot(kind="scatter",x=['departure_longitude'],y=['departure_latitude'],figsize=(15,10))
ax.set_xlabel("Longitude")
ax.set_ylabel("Latitude")
```

#### Let's see how many unique destinations there

```python
print("The destinations departure options to choose from "+str(df['departure_id'].nunique())+" and destinations to return "+str(df['return_id'].nunique()))
```

#### Turning database into Time Series

```python
time_df = df
time_df[['departure_f']] = time_df.departure
time_df.departure_f = pd.to_datetime(time_df.departure_f,format="%Y-%m-%dT%H:%M:%S")
time_df.index=time_df.departure_f
time_df.drop(['departure_f'], axis = 1,inplace=True)
time_df
```

#### Counting uses for every month

```python
monthly_count = time_df[['distance (m)']].resample('1M').count()
monthly_count['total_dist'] = time_df[['distance (m)']].resample('1M').sum()
plt.plot(monthly_count.index,monthly_count['distance (m)'])
```

*As a non-finnish person it seemed weird, but the zero values are due to snowfall, where u can't take bikes for ride*

### Plotting popular departure desitnations with longitude and latitude

```python
dept_df = df[['departure','departure_latitude','departure_longitude']]
dept_df = dept_df.dropna()
count_depart = dept_df.groupby(['departure_latitude','departure_longitude'])[['departure']].count()
count_depart.reset_index(inplace=True)
count_depart.plot(kind="scatter",x="departure_longitude",y="departure_latitude",s=count_depart["departure"]/2000,figsize=(15,10),label="Populat Departure Destination",c=count_depart["departure"]/1000,cmap=plt.get_cmap("jet"),colorbar=True)
```
