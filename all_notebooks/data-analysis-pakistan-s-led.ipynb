```python
main_data = pd.read_csv("/kaggle/input/pakistans-largest-ecommerce-dataset/Pakistan Largest Ecommerce Dataset.csv", 
                        parse_dates=["created_at", "Working Date"], low_memory=False)
```

```python
status_col = main_data.status
status_col.isna().sum()
```

```python
status_col = status_col.dropna()
status_list = list(set(status_col))
print(status_list)
```

```python
status_list = status_list[1:]
print(status_list)
```

```python
status_values = []
for s in status_list:
    val = main_data[main_data.status == s].status.count()
    status_values.append(val)


status_info = np.array([status_list, status_values])
```

```python
print(status_info)
```

```python
for i in range(len(status_list)):
    print(status_info[0][i], ' = ', status_info[1][i])
```

```python
import matplotlib.pyplot as plt
plt.barh(status_list,status_values)
plt.title('Status of Orders against total Orders', fontsize = 14)
plt.xlabel('Order Counts', fontsize = 14)
plt.ylabel('Order Status', fontsize = 14)
plt.grid(True)
plt.autoscale(tight=True)
plt.show()
```
