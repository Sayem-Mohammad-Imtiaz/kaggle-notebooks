```python
df = pd.read_csv("/kaggle/input/olympic-games-2021-medals/Tokyo 2021 dataset.csv")
df["Team/NOC"] = df["Team/NOC"].replace({'ROC': 'Russian Federation'})
df["NOCCode"] = df["NOCCode"].replace({'ROC': 'RUS'})
```

```python
import chart_studio.plotly as py
import plotly.offline as po
import plotly.graph_objs as go
```

```python
data = dict(type='choropleth', 
            locations = df['NOCCode'], 
            z = df['Gold Medal'], 
            text = df['Team/NOC'],
            colorscale="peach")

layout = dict(title = '<b>Gold medals across the world</b>', 
              geo = dict(showframe = False, 
                       projection = {'type':'orthographic'}, 
                       showlakes = True, lakecolor="blue",
                       showocean=True, oceancolor="LightBlue",
                       showland=True, landcolor="LightGreen"))


fig = go.Figure(data = [data], layout = layout, )
fig.update_layout(title_x=0.5, width=700, height=700, font=dict(family="Times New Roman", size=18), font_color="#000000", title_font_color="#FF9833")
po.iplot(fig)
```
