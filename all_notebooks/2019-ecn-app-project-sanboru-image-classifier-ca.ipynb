 3.Feature Engineering

3.1 Feature Selection

```python
features = ["sample_url", "tags"]
df_X = df_raw[features]
df_X.columns
```

```python
df_X.head(samples)
```

3.2 Tags extraction

```python
import re
tag = []

for i in df_X.tags:
    tokens = re.split("[ ]",i)
    for token in tokens:
        if token not in tag:
            tag.append(token)
print("There are", len(tag), "different tags")

tag[:10]
```

```python
chose = []
dic = {}
for i in df_X.tags:
    tokens = re.split("[ ]",i)
    for token in tokens:
        if token in ['1girl', 'bag','black_hair','blush','bob_cut']:
            chose.append(token)
for j in chose:
     dic[j] = dic.get(j,0)+1      
print("5 tags occurences:",dic)
```

```python
dict = {}
list = []
for i in df_X.tags:
    tokens = re.split("[ ]",i)
    for token in tokens:
        list.append(token)
for i in list:
    dict[i] = dict.get(i,0)+1

item = sorted(dict.items(), key = lambda x:x[1],reverse = True)
print("10 top tags:")
for i in range(0,10):
    print(item[i])
    
```
