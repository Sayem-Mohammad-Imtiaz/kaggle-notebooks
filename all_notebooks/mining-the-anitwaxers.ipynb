# Introduction
The Anti Vaccination campaign is one of the significant challenges in Public Health. Vaccines are one of the notable contributions of science to humanity. When living in a pandemic affected era, the majority of people realizes how vital is the vaccine. At the same, some people believe vaccines are against their personal beliefs or are part of conspiracy theories. Accessibility to technology and the social network made it easy for anti-vaccine propaganda generators to preach their philosophy. Surprisingly they made some impressions, and some of the developing countries and developed countries suffered a lot. Measles outbreak happened in 2020 in the USA is one of the recent examples. Alec Helyar, a Kaggler and his collaborators, created a data-set from various social media sites. The current work is an attempt to explore the language and the medium of AntiVaxx propaganda. 

```python

%matplotlib inline
import os
import warnings
warnings.simplefilter(action='ignore')

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import sklearn as sl
```

```python
text_data = pd.read_csv("../input/facebook-antivaccination-dataset/posts_full.csv")
```

**Since the data don't have many details, we will explore the attributes by the name.**

```python
text_data.article_host.value_counts().nlargest(15).plot(kind='bar',
                                                       figsize=(15,5),
                                                       alpha=0.9,
                                                        rot=35,
                                                       title="Top 15 AntiVaxx Content Posting Sources")
```
