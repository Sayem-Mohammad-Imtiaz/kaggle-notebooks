3. Obtén una lista con los atributos cuantitativos y otra lista con los atributos cualitativos.

```python
lista =[]
for i in df.iloc[0].values:
    try:
        if type(float(i))==float:
            lista.append("cuantitativo")
    except:
        lista.append("cualitativo")
lista
```

```python
for i,j in zip(lista,df.columns):
    if i == "cuantitativo":
        df[j] = df[j].astype("float")
    else:
        df[j]= df[j].astype("category")
```

6. Obtén el rango, varianza, desviación estandar para los atributos cualitativos.

```python
for i in range(4,120):
    rango = np.max(df[df.columns[i]].values) - np.min(df[df.columns[i]].values)
    varianza = np.var(df[df.columns[i]].values)
    destandar = np.std(df[df.columns[i]].values)
    print(df[df.columns[i]].name,'\n R = ',rango,'\n S² = ', varianza, '\n S = ',destandar)
```
