### importing libraries and read Date set 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv('/kaggle/input/aus-real-estate-sales-march-2019-to-april-2020/aus-property-sales-sep2018-april2020.csv')
df.head()
```

#### exploring data type and null values 

```python
df.info()
```

#### we need to make some cleaning

##### date_sold need to converted to Datetime

##### Price have null values need to fill


```python
df['date_sold'] = pd.to_datetime(df['date_sold'])
```

```python
df['price'].fillna((df['price'].mean()), inplace=True)
```

```python
df.info()
```

##### now if we rmoved the Nan values remaining , it will not affect the resualt

```python
df.dropna().head()
```

```python
df['city_name'].value_counts()
```

##### its clear that the state is the same city name , so i will remove the 

```python
df.drop("state", axis=1, inplace=True)
```

##### count the numbers of bed rooms 

```python
df['bedrooms'].value_counts()
```

```python
df['property_type'].value_counts()
```

##### present the top of 20 suburb 

```python
df['suburb'].value_counts().head(20)
```

```python
count_city = pd.DataFrame(df.city_name.value_counts())
labels_city= count_city['city_name'].index 
values_city =count_city['city_name'] 

count_property = pd.DataFrame(df.property_type.value_counts())
labels_property= count_property['property_type'].index 
values_property =count_property['property_type'] 
```

```python
fig = make_subplots(rows=1, cols=2, specs=[[{'type':'domain'}, {'type':'domain'}]])
fig.add_trace(go.Pie(labels=labels_city, values=values_city, name="City name "),
              1, 1)
fig.add_trace(go.Pie(labels=labels_property , values= values_property  , name="Property_type"),
              1, 2)

# Use `hole` to create a donut-like pie chart
fig.update_traces(hole=.4, hoverinfo="label+percent+name")

fig.update_layout(
    title_text="property and city name chart ",
    # Add annotations in the center of the donut pies.
    annotations=[dict(text='City ', x=0.18, y=0.5, font_size=20, showarrow=False),
                 dict(text='property', x=0.82, y=0.5, font_size=20, showarrow=True)])
fig.show()
```
