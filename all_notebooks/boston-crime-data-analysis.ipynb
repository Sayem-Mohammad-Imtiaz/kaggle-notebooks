```python
data.SHOOTING = data.SHOOTING.fillna("N")
sns.countplot(data.SHOOTING,)
```

1. - 枪击犯罪在所有犯罪案件中占比是极低的

```python
plt.figure(figsize=(16,6))
sns.countplot(data.OFFENSE_CODE_GROUP)
plt.xticks(rotation=-90)
```

- 上图很容易看出不同犯罪分组发生的次数

```python
plt.figure(figsize=(16,6))
sns.countplot(data.DISTRICT)
plt.xticks(rotation=-90)
```

**由上图可知，犯罪发生的主要区域为B2区**

```python
# 下面进行统计不同年月发生犯罪的频率差异
fig = plt.figure(figsize=(16,12))
sns.countplot(data.YEAR,ax=fig.add_subplot(221))
sns.countplot(data.MONTH,ax=fig.add_subplot(222))
sns.countplot("MONTH",data=data,hue="YEAR",ax=fig.add_subplot(212))
```

1. 从年份来分析，2016、2017年犯罪次数上升，而2018年的时候犯罪似乎下降了，其实并不是这样，因为统计原因，2015年前几个月缺失数据，2018年后几个月没有数据，所以图表中展现了犯罪率随年份变化的假象。考虑数据缺失的情况，四年的犯罪率并没有多大的波动。
1. 从月份来分析，应按照不同年份单独解读图表。考虑数据的完整性，我们只看2016、2017的数据。犯罪率随着月份的变化呈现了规律性的变化，7-9月犯罪率最高的时间。

*犯罪率是否在一天的不同时段存在差异呢？*

```python
plt.figure(figsize=(16,8))
sns.countplot(data.HOUR)
```

**很显然，犯罪频率在每天不同的时段展现了很强的差异，白天比晚上更容易发生犯罪**

```python
# 我们对地点绘制成散点图,部分坐标数据可能有误，故设置x、y轴坐标范围
plt.figure(figsize=(12,12))
sns.scatterplot(x="Lat",y="Long",data=data)
plt.axis([42.2,42.45,-71.2,-70.95])
```

- 我们可以看出犯罪发生的位置分布是有特点的，集中分布在几个特定的区域
