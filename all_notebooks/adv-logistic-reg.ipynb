```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
ad_data = pd.read_csv('/kaggle/input/advertising/advertising.csv')
print(ad_data.info())
#print(ad_data.describe())
```

```python
ad_data['hour'] = ad_data['Timestamp'].apply(lambda x: pd.Timestamp(x).hour) # extrag ora din coloana de timp
ad_data.drop(['Ad Topic Line', 'City', 'Country', 'Timestamp'], axis=1, inplace=True)
print(ad_data.head())
```

```python
# check for incomplete data
sns.heatmap(ad_data.isnull())
plt.show()
```

```python
X = ad_data.drop('Clicked on Ad', axis=1)
y = ad_data['Clicked on Ad']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
logmod = LogisticRegression(max_iter=10000)
logmod.fit(X_train, y_train)
predictions = logmod.predict(X_test)
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```
