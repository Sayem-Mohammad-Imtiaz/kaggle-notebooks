## Linear Regressions Model with Stats

```python
sns.jointplot(x="year",y="population",data = df,kind="reg");
sns.jointplot(x="density",y="population",data = df,kind="reg");
```

```python
from sklearn.model_selection import train_test_split,cross_val_score,cross_val_predict
```

```python
X = df.drop(columns=["population","change_perc","net_change","urban_perc"],axis = 1)
y = df["population"]

import statsmodels.api as sms
lm = sms.OLS(y,X)
model = lm.fit()
model.summary()
```

```python
model.params
```

```python
model.conf_int()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20 ,random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model=lm.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred =model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
model.score(X,y)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
