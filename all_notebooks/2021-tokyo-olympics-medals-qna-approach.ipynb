## 1) What are the total number of countries that obtain a gold medal at least?

## 2) What are total number of each gold, silver and bronze medals awarded? Are they equal?
I heard that there is a high jump event where two athletes from Italy and Bahrain respectively share a gold medal together. So I expect the total gold medals to be 2 more than the total silver medals.

```python
total_gold = df[df["Gold Medal"]>=1]["Gold Medal"].sum()
total_silver = df[df["Silver Medal"]>=1]["Silver Medal"].sum()
total_bronze = df[df["Bronze Medal"]>=1]["Bronze Medal"].sum()
print('Total Gold Medals',total_gold)
print('Total Silver Medals',total_silver)
print('Total Bronze Medals',total_bronze)
```

Yup, that was right! And the bronze medals are quite higher than the gold and silver medals. I believe it is because some of the events do not have a bronze medal match such as boxing.
