## Line Charts (Çizgi Grafikler) 

Line Charts Example : Citation and Teaching vs World Rank of Top 100 University
(Çizgi Grafik Örneği: Atıf ve Öğretim - İlk 100 Üniversitenin Dünya Sıralaması)

* import graph _objs as go 
* Creating traces (
    * x = x axis  (x ekseni)
    * y = y axis  (y ekseni)
    * mode = types of plot like marker , line or line + markers 
    * name = name of the plots
    * marker = marker is used with dictionary 
        * color = color of line. It takes RGB and opacity(alpha)  (renk ve saydamlık)
    * text = the hover text (hover is curser)  (grafiğin üzerine geldiğimizde ne tür yazı görmek istiyoruz)
    
* data = is a list that we add traces into it 
* layout = it is dictionary 
    * title = title of layout 
    * x axis = it is dictionary 
        * title = label of x axis 
        * ticklen = length of x axis ticks 
        * zeroline = showing zero line or not 
* fig = it includes data and layout 
* implot() = plots the figure(fig) that is created by data and layout 

```python
# prepare data frame
df = timesData.iloc[:100,:]

# import graph objecets as  "go"
# "go" ile kütüphaneyi çağırdık
import plotly.graph_objs as go 


trace1 = go.Scatter(
                    x = df.world_rank,      # (x eksenine word_rank eklendi )
                    y = df.citations,       # (y eksenine citations eklendi ) 
                    mode = "lines",         # lines => düz çizgiler 
                    name = "citations",     # name => çizginin adı
                    marker = dict(color = "rgba(16,112,2,0.8)"),   # çizginin rengi ve saydamlığı
                    text = df.university_name)  # çizginin üzerine geldiğimiz zaman üniversitenin adı yazacak


trace2 = go.Scatter(                 # Scatter kullandık 
                    x = df.world_rank,
                    y = df.teaching,
                    mode = "lines+markers",   # noktalı çizgiler oluşturduk
                    name = "teaching",
                    marker = dict(color = "rgba(80,26,80,0.8)"),
                    text = df.university_name)


data = [trace1,trace2]   # iki scatteri birleştirdik , liste yapmış olduk 

layout = dict(title = "Citation and Teaching vs World Rank of Top 100 Universities", # tablonun başlığını belirledik
             xaxis = dict(title= "World Rank",ticklen=5,zeroline= False)   # x ekseni bilgisini ekledik ,ticklen = kalınlık 
             )

fig = dict(data = data ,layout = layout)
iplot(fig)
```

## Scatter Plot (Dağılım Grafiği)

Scatter Example : Citation vs world rank of top 100 universities with 2014, 2015 and 2016 years

* Import graph_objs as go
* Creating traces
    * x = x axis
    * y = y axis
    * mode = type of plot like marker, line or line + markers
    * name = name of the plots
    * marker = marker is used with dictionary.
        * color = color of lines. It takes RGB (red, green, blue) and opacity (alpha)
    * text = The hover text (hover is curser)
* data = is a list that we add traces into it
* layout = it is dictionary.
    * title = title of layout
    * x axis = it is dictionary
        * title = label of x axis
        * ticklen = length of x axis ticks
        * zeroline = showing zero line or not
    * y axis = it is dictionary and same with x axis
* fig = it includes data and layout
* iplot() = plots the figure(fig) that is created by data and layout

```python
# prepare data frame

df2014 = timesData[timesData.year == 2014].iloc[:100,:]  # 2014 yılı için ilk 100 üniversite alındı
df2015 = timesData[timesData.year == 2015].iloc[:100,:]  # 2015 yılı için ilk 100 üniversite alındı
df2016 = timesData[timesData.year == 2016].iloc[:100,:]  # 2016 yılı için ilk 100 üniversite alındı

#import graph objects as "go"
#tekrardan import edildi 
import plotly.graph_objs as go

trace1 = go.Scatter(x = df2014.world_rank,y = df2014.citations,mode="markers",name="2014",  # Scatter  kullanıldı ,
                    marker = dict(color ="rgba(255,128,255,0.8)"),text = df2014.university_name)  #  sadece noktalı gösterim olduğu için markers kullanıldı

trace2 = go.Scatter(x = df2015.world_rank,y = df2015.citations,mode="markers",name="2015",
                    marker = dict(color ="rgba(255,128,2,0.8)"),text = df2015.university_name)

trace3 = go.Scatter(x = df2016.world_rank,y = df2016.citations,mode="markers",name="2016",
                    marker = dict(color ="rgba(0,255,200,0.8)"),text = df2016.university_name)

                    
data = [trace1,trace2,trace3]   # Scatterler liste haline getirildi
                    
                    
layout = dict(title = "'Citation vs world rank of top 100 universities with 2014, 2015 and 2016 years",
                     xaxis = dict(title = "World Rank", ticklen=5 ,zeroline= False),  # y ekseni için yazı
                     yaxis = dict(title = "Citation",ticklen=5 ,zeroline= False)      # x ekseni için yazı
             )
    
                    
fig = dict(data = data , layout = layout)
iplot(fig)
```

## Bar Charts  (Çubuk Grafikler)

        * line = It is dictionary. line between bars
            * color = line color around bars

* data = is a list that we add traces into it
* layout = it is dictionary.
    * barmode = bar mode of bars like grouped
* fig = it includes data and layout
* iplot() = plots the figure(fig) that is created by data and layout

```python
df2014 = timesData[timesData.year == 2014].iloc[:3,:]
df2014
```

```python
data2014 = timesData[timesData.year == 2014].iloc[:3,:]  # ilk 3 üniversiteyi aldık

import plotly.graph_objs as go

trace1 = go.Bar(x = df2014.university_name,
               y = df2014.citations,
               name = "citations", # kutunun adı
               marker = dict(color="rgba(255,174,255,0.5)",  # kutunun rengi
               line = dict(color = "rgb(0,0,0)",width=1.5)),
               text = df2014.country)


trace2 = go.Bar(x = df2014.university_name,  # grafik tipi bar olduğu için mode özelliği kullanılmadı
               y = df2014.teaching,
               name = "teaching",    
               marker = dict(color="rgba(255,255,128,0.5)",
               line = dict(color = "rgb(0,0,0)",width=1.5)), #barların çerçevesinin rengini ve kalınlığını belirliyor
               text = df2014.country)  # grafiğin üstüne geldiğimiz zaman özellikle üniversitenin hangi ülkede olduğunu belirten parça 

data = [trace1,trace2]

layout = go.Layout(barmode = "group")   # barmode ile kutuların grupça durmasını sağladık
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

Third Bar Charts Example: Horizontal bar charts. (style3) Citation vs income for universities


* Import graph_objs as go and importing tools
    * Tools: used for subplots
* Creating trace1
    * bar: bar plot
        * x = x axis
        * y = y axis
        * marker
            * color: color of bars
            * line: bar line color and width
        * name: name of bar
        * orientation: orientation like horizontal
            * creating trace2
            * scatter: scatter plot
                * x = x axis
                * y = y axis
                * mode: scatter type line line + markers or only markers
                * line: properties of line
                    * color: color of line
                * name: name of scatter plot
* layout: axis, legend, margin, paper and plot properties 

```python
from plotly import tools

df2016 = timesData[timesData.year == 2016].iloc[:7,:]

y_saving = [each for each in df2016.research]
y_net_worth = [float(each) for each in df2016.income]
x_saving = [each for each in df2016.university_name]
x_net_worth = [each for each in df2016.university_name]

trace0 = go.Bar(
                x = y_saving,
                y = x_saving,
                marker = dict(color="rgba(171,50,96,0.6)",
                             line = dict(color="rgba(171,50,96,1.0)",
                                        width = 1)),
    name = "research",
    orientation = "h", ) 

trace1 = go.Scatter(
                    x = y_net_worth,
                    y = x_net_worth,
                    mode = "lines+markers",
                    line=dict(color="rgb(63,72,205)"),
                    name = "income")
layout = dict(
                title="kflgmlfdgdfb",
                yaxis = dict(showticklabels=True,domain=[0,0.85]),
                yaxis2 = dict(showline=True,showticklabels=False,linecolor="rgba(102,102,102,0.8)",linewidth=2,domain=[0,0.85]),
    
                xaxis = dict(zeroline=False,showline=False,showticklabels=True,showgrid=True,domain=[0,0.42]),
                xaxis2 = dict(zeroline=False,showline=False,showticklabels=True,showgrid=True,domain=[0.47,1],side="top",dtick=25),
    
                legend = dict(x=0.029,y=1.038,font= dict(size = 10)),
        
                margin = dict(l=200,r=20,t=70,b=70),
                paper_bgcolor = "rgb(248,248,255)",
                plot_bgcolor = "rgb(248,248,255)",)


annotations = []

y_s = np.round(y_saving,decimals=2)
y_nw = np.rint(y_net_worth)

for ydn, yd, xd in zip(y_nw,y_s,x_saving):
    
    annotations.append(dict(xref="x2",yref="y2",y=xd,x=ydn - 4,text = "(:,)".format(ydn),font=dict(family="Arial",size=12,color="rgb(63,72,204)"),
                           showarrow = False))
    
    annotations.append(dict(xref='x1', yref='y1', y=xd, x=yd + 3,text=str(yd),font=dict(family='Arial', size=12,color='rgb(171, 50, 96)'),
                            showarrow=False))
    

    
layout["annotations"] = annotations

fig = tools.make_subplots(rows=1,cols=2,specs=[[{},{}]],shared_xaxes=True,
                          shared_yaxes = False, vertical_spacing=0.001)


fig.append_trace(trace0,1,1)
fig.append_trace(trace1,1,2)

fig["layout"].update(layout)
iplot(fig)
```

## Pie Charts (Pasta grafikler)

Pie Charts Example : Students rate of top 7 university in 2016
(Pasta Grafikleri Örneği: 2016'da en iyi 7 üniversitenin öğrenci oranı)

* fig: create figures (rakamlar oluştur)
    * data: plot type
        * values: values of plot
        * labels: labels of plot
        * name: name of plots
        * hoverinfo: information in hover
        * hole: hole width
        * type: plot type like pie
    * layout: layout of plot
        * title: title of layout
        * annotations: font, showarrow, text, x, y


```python
# data preparation

df2016 = timesData[timesData.year == 2016].iloc[:7,:]  #ilk 7 veriyi aldık

pie1 = df2016.num_students

pie1_list = [float(each.replace(",",".")) for each in df2016.num_students]
# num_students değerleri virgül ile ondalıkları gösterildiği için bunları noktaya çeviriyoruz 
# değeri normalde object türünde , floata çevirdik
# str(2,4) => str(2.4) => float(2.4)

labels = df2016.university_name # ilk 7 üniversitenin sadece isimleri alındı

fig = {
    
    "data":[
        {
            "values" : pie1_list,  # düzenlediğimiz öğrenci numaraları
            "labels" : labels,     # üniversite isimleri
            "domain" : {"x" : [0,0.5]}, # ? 
            "name" : "Number Of Students Rate",
            "hoverinfo" : "label+percent",  # fare ile üstüne gelinen bölüm için ünversite adı ve yüzdelik ifadesi hakkında bilgi veriyor
            "hole" : 0.3,    # çember içindeki beyaz bölüm , geri kalan bölüm 0.7 ye denk geliyor
            "type" : "pie",  # grafik türü belirlendi
                             # grafik türü num_students değerlerini topladı ve % lik olarak oranladı
            
        },
    ],
    
    "layout" : {
        "title" : "Universities Number of Students rates",  # genel başlık 
        "annotations" : [
            {
                "font" : {"size" : 20},
                "showarrow" : False,
                "text": "Number of Students",  # alt başlık ; büyüklüğü , x ve y konumu belirlendi
                "x" : 0.3,
                "y" : 1.1,
            }
        ]
    }
}

iplot(fig)
```

## Histogram 

Lets look at histogram of students-staff ratio in 2011 and 2012 years.

* trace1 = first histogram
    * x = x axis
    * y = y axis
    * opacity = opacity of histogram
    * name = name of legend
    * marker = color of histogram
* trace2 = second histogram
* layout = layout
    * barmode = mode of histogram like overlay. Also you can change it with stack

```python
# prepara data 

x2011 = timesData.student_staff_ratio[timesData.year == 2011]  # 2011 ve 2012 yıllarını aldık
x2012 = timesData.student_staff_ratio[timesData.year == 2012]

trace1 = go.Histogram(  # grafik türünü belirledik
    x = x2011,
    opacity = 0.75, # saydamlık 
    name = "2011",
    marker = dict(color = "rgba(171,50,96,0.6)"))

trace2 = go.Histogram(
    x = x2012,
    opacity = 0.75,
    name = "2012",
    marker = dict(color = "rgba(12,50,196,0.6)"))

data = [trace1 , trace2]

layout = go.Layout(barmode = "overlay",  # overlay ile histogramları üst üste bindirildi
                  title = "students-staff ratio in 2011 and  2012",
                  xaxis = dict(title = "students-staff ratio"),
                  yaxis = dict(title = "Count"),
                  )

fig = go.Figure(data = data , layout = layout)
iplot(fig)
```

## Box Plots  (Kutu Grafikleri)

Box Plots

* Median (50th percentile) = middle value of the data set. Sort and take the data in the middle. It is also called 50% percentile that is 50% of data are   less that median(50th quartile)(quartile)

    * 25th percentile = quartile 1 (Q1) that is lower quartile
    * 75th percentile = quartile 3 (Q3) that is higher quartile
    * height of box = IQR = interquartile range = Q3-Q1
    * Whiskers = 1.5 * IQR from the Q1 and Q3
    * Outliers = being more than 1.5*IQR away from median commonly.

* trace = box
    * y = data we want to visualize with box plot
    * marker = color

```python
# data preparation 

x2015 = timesData[timesData.year == 2015]

trace00 = go.Box(
    y = x2015.total_score,  # 2015 senesinin total_score verilerini aldık
    name = "total score of universities in 2015",
    marker = dict(
        color = "rgb(12,12,140)",
    )
)

trace11 = go.Box(
    y = x2015.research, # 2015 senesinin research verilerini aldık
    name = "research of universities in 2015",
    marker = dict(
        color = "rgb(12,128,128)",
    )
)
data = [trace00,trace11]

iplot(data)

```

## Scatter Matrix Plots (Dağılım Matrisi Grafikleri)


Scatter Matrix = it helps us to see covariance and relation between more than 2 features
(Dağılım Matrisi = 2'den fazla özellik arasındaki kovaryansı ve ilişkiyi görmemize yardımcı olur)

* import figure factory as ff
* create_scatterplotmatrix = creates scatter plot (dağılım grafiği oluşturur)
    * data2015 = prepared data. It includes research, international and total scores with index from 1 to 401
    * (data2015 = hazırlanmış veriler. 1'den 401'e kadar indeksli araştırma, uluslararası ve toplam puanları içerir.)
    * colormap = color map of scatter plot  (dağılım grafiğinin renk haritası)
    * colormap_type = color type of scatter plot  (dağılım grafiğinin renk türü)
    * height and weight (yükseklik ve ağırlık)

```python
# import figure factory

import plotly.figure_factory as ff 

#prepare data

dataframe = timesData[timesData.year == 2015]
data2015 = dataframe.loc[:,["research","international","total_score"]]  # sadece bu sütunlar alındı
data2015["index"] = np.arange(1,len(data2015) + 1)  # eleman sayısı kadar index değerimiz oluyor

#scatter matrix

fig = ff.create_scatterplotmatrix(data2015, diag = "box", # box plot 
                                  index = "index",
                                  colormap = "Portland",  # maviden kırmızıya giden renk paleti
                                 colormap_type = "cat",   # dağılım grafiğinin renk türü
                                 height = 700 , width = 700)

iplot(fig)
```

## Inset Plot (Ekli Küçük Resim Grafiği)

Inset Matrix = 2 plots are in one frame  (Inset Matrix = 2 grafik tek çerçevede)



```python
# firs line plot

trace1 = go.Scatter(
    x = dataframe.world_rank,
    y = dataframe.teaching,
    name = "teaching",
    marker = dict(color = "rgba(16,112,2,0.8)"),
)

#second line plot

trace2 = go.Scatter(
    x = dataframe.world_rank,
    y = dataframe.income,
    xaxis = "x2",
    yaxis = "y2",
    name = "income",
    marker = dict(color = "rgba(160,112,20,0.8)"),
)

data = [trace1,trace2]

layout = go.Layout(  
    xaxis2 = dict(
        domain = [0.4,1],  # küçük grafiğin kordinatları ile alakalı
        anchor = "y2",
    ),
    yaxis2=dict(
        domain=[0.8, 1],   # küçük grafiğin kordinatları ile alakalı
        anchor='x2',
    ),
    
    title = "Income and Teaching vs World Rank of Universities"
)
fig = go.Figure(data = data,layout = layout)
iplot(fig)


```

## 3D Scatter Plot With Colorscaling (Renk Ölçekleme ile 3D Dağılım Grafiği)

3D Scatter: Sometimes 2D is not enough to understand data. Therefore adding one more dimension increase the intelligibility of the data. Even we will add color that is actually 4th dimension.

(3D Dağılım: Bazen 2D, verileri anlamak için yeterli değildir. Bu nedenle, bir boyut daha eklemek verilerin anlaşılırlığını artırır. Hatta aslında 4. boyut olan rengi de ekleyeceğiz.)

* go.Scatter3d: create 3d scatter plot  (3d dağılım grafiği oluştur)
* x,y,z: axis of plots
* mode: market that is scatter
* size: marker size
* color: axis of colorscale (renkli ölçek ekseni)
* colorscale: actually it is 4th dimension (aslında 4. boyut)

```python
# create trace 1 that is 3D scatter

trace1 = go.Scatter3d(
    x = dataframe.world_rank,
    y = dataframe.research,
    z = dataframe.citations,
    mode = "markers",  # nokta gösterimli grafik
    marker = dict(color = "rgb(255,0,0)")
)
data = [trace1]
layout = go.Layout(
    margin = dict(  # kenardan bırakılan boşluklar
    l = 0,
    r = 0,
    b = 0,
    t = 0,
    )
)

fig = go.Figure(data = data , layout = layout)
iplot(fig)
```

## Multiple Subplots  (Çoklu Alt Noktalar)

Multiple Subplots: While comparing more than one features, multiple subplots can be useful.

Çoklu Alt Grafikler: Birden fazla özelliği karşılaştırırken, birden çok alt grafik yararlı olabilir.


```python

trace1  = go.Scatter(
    x = dataframe.world_rank,
    y = dataframe.research,
    name = "research")

trace2 = go.Scatter(
    x = dataframe.world_rank,
    y = dataframe.citations,
    xaxis = "x2",
    yaxis = "y2",
    name = "citations")

trace3 = go.Scatter(
    x = dataframe.world_rank,
    y = dataframe.income,
    xaxis = "x3",
    yaxis = "y3",
    name = "income")

trace4 = go.Scatter(
    x = dataframe.world_rank,
    y = dataframe.total_score,
    xaxis = "x4",
    yaxis = "y4",
    name = "total_score")

data = [trace1,trace2,trace3,trace4]

layout = go.Layout(
    xaxis = dict(
        domain = [0,0.45]
    ),
    yaxis = dict(
        domain = [0,0.45]
    ),
    xaxis2 = dict(
        domain = [0.55,1]
    ),
    xaxis3 = dict(
        domain = [0,0.45],
        anchor = "y3"
    ),
    xaxis4 = dict(
        domain = [0.55,1],
        anchor = "y4"
    ),
    yaxis2 = dict(
        domain = [0,0.45],
        anchor = "x2"
    ),
    yaxis3 = dict(
        domain = [0.55,1] 
    ),
    yaxis4 = dict(
        domain = [0.55,1],
        anchor = "x4"
    ),
    title = "Research, citation, income and total score VS World Rank of Universities"
)
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```
