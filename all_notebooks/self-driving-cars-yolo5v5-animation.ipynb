# Create animation

```python
from matplotlib import animation, rc
rc('animation', html='jshtml')
```

```python
def create_animation(ims):
    fig=plt.figure(figsize=(12,12))
    plt.axis('off')
    im=plt.imshow(cv2.cvtColor(ims[0],cv2.COLOR_BGR2RGB))
    
    def animate_func(i):
        im.set_array(cv2.cvtColor(ims[i],cv2.COLOR_BGR2RGB))
        return [im]

    return animation.FuncAnimation(fig, animate_func, frames=len(ims), interval=1000//4)
```

```python
imgdir ='/kaggle/working/yolov5/runs/detect/exp/' 
```

```python
paths0=[]
for dirname, _, filenames in os.walk(imgdir):
    for filename in filenames:
        paths0+=[os.path.join(dirname, filename)]
paths0[0:5]
```

```python
paths1=[]
for item in paths0:
    if item[-4:]=='.jpg':
        paths1+=[item]
paths1[0:5]
```

```python
order=[]
for item in paths1:
    order+=[int(item[:-6].split('/')[-1])]
paths2=pd.DataFrame(paths1)
paths2[1]=order
paths2.columns=['path','int']
paths2=paths2.sort_values('int')
paths3=paths2['path'].tolist()
paths3[0:5]
```

```python
images0=[]
for i in tqdm(range(0,len(paths3),1)):
    images0+=[cv2.imread(paths3[i])]
```
