# Training the Model 

```python
#Making the object of the LinearRegression

linreg = LinearRegression()
```

```python
# Splitting the dataset into the independent and dependent features

X = data.iloc[:,:-1] # Independent Features
Y = data.iloc[:,-1] # Dependent Features
```

```python
# Splitting the data into train and test from the dataset

X_train , X_test , Y_train , Y_test = train_test_split(X,Y,random_state = 0, test_size = 0.2 )
```

```python
# Checking the size of train and test data

print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

```python
# Training the Model

linreg.fit(X_train,Y_train)
```

```python
# Testing the Model

Y_pred = linreg.predict(X_test)
```

```python
# Comparing the Actual Price and the Predicted Price

model_result = pd.DataFrame({"Actual_Price" : Y_test, "Predicted_Price" : Y_pred})
```

```python
model_result.head()
```

```python
# Checking the accuracy of the model using the r2_score

r2 = r2_score(Y_test , Y_pred)
r2
```

# Thus we can say that our model is 100% accurate

# Now predicting the house price using the user input

```python
area = int(input("Enter the Area of the House: "))
garage = int(input("Enter the number of garage: "))
firePlace = int(input("Enter the number of fire place: "))
bathroom = int(input("Enter the number of bathroom: "))
whiteMarble = int(input("Enter 1 for white marble else 0: "))
blackMarble = int(input("Enter 1 for black marble else 0: "))
indianMarble = int(input("Enter 1 for indian marble else 0: "))
floors = int(input("Enter the number of floors in the house: "))
city = int(input("Enter the larger value for the good city range(1-3): "))
solar = int(input("Enter 1 if you want solar system else 0: "))
electric = int(input("Enter 1 if you need electric work to be done else 0: "))
fiber = int(input("Enter 1 or 0 for the electric fibre: "))
glassDoors = int(input("Enter 1 if you want glass doors else 0: "))
swimingPool = int(input("Enter 1 if you want a swiming pool else 0: "))
garden = int(input("Enter 1 if you want garden in the house else 0: "))

values = [area,garage,firePlace,bathroom,whiteMarble,blackMarble,indianMarble,floors,
          city,solar,electric,fiber,glassDoors,swimingPool,garden]

col = ['area','garage','firePlace','bathroom','whiteMarble','blackMarble',
                      'indianMarble','floors','city','solar','electric','fibre','glassDoors'
                     ,'swimingPool','garden']

user_input = pd.DataFrame(data = [values],
                         columns = col)
```

```python
# Predicting the price according to the user input

user_predicted_price = linreg.predict(user_input)
```

```python
user_predicted_price = float(user_predicted_price)
formatted_price = "{:.2f}".format(user_predicted_price)
print("According to the User Input ${} is the price of the house".format(formatted_price))
```
