```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import validation_curve
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../input/internationalmigration/international-migration-December-2020-estimated-migration-by-age-sex.csv")
```

```python
data.shape # print the dimension or shape of train data
```

```python
data.head()
```

```python
X = data.drop(data.columns[[0, 1, 2]], axis=1).values
X = X[:,:-1]
y = data.iloc[:, -1].values
print(X[0])
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 0] = le.fit_transform(X[:, 0])
#lw = LabelEncoder()
X[:, 1] = le.fit_transform(X[:, 1])
X[:, 2] = le.fit_transform(X[:, 2])
y[:] = le.fit_transform(y[:])

print(X[0])
print(y[0])
```

```python
y=y.astype('int')
```

```python
print("X:", X.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size=0.5,test_size = 0.1,random_state = 10)
```

```python
X_train
```

```python
y_train
```

```python
import sklearn
# model with optimal hyperparameters

# model
model = SVC(C=10, gamma=0.001, kernel="rbf")

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# metrics
print("accuracy", sklearn.metrics.accuracy_score(y_test, y_pred), "\n")
print(sklearn.metrics.confusion_matrix(y_test, y_pred), "\n")
print(sklearn.metrics.classification_report(y_test, y_pred))
```
