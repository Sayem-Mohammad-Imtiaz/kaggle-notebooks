```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
```

```python
data = pd.read_csv("/kaggle/input/kidney-kronicle/kidney_disease.csv")
```

```python
y_data = data["classification"]
data = data.drop(["id", "classification"], axis = 1)
```

```python
y_data[(y_data=="ckd")]=1
y_data[(y_data=="ckd\t")]=1
y_data[(y_data=="notckd")]=0
```

```python
cols = ["rbc", "pc", "pcc", "ba", "htn", "dm", "cad", "appet", "pe", "ane"]
data_onehot = pd.get_dummies(data[cols])
data = pd.concat([data, data_onehot], axis=1)
data = data.drop(cols, axis = 1)
```

```python
cols = ["pcv", "wc", "rc"]
data = data.drop(cols , axis = 1)
```

```python
from sklearn.impute import SimpleImputer
# cols_with_missing = [col for col in X_train.columns if X_train[col].isnull().any()]

my_imputer = SimpleImputer( strategy='mean')
data = pd.DataFrame(my_imputer.fit_transform(data))
```

```python
data = np.array(data, dtype = 'float32')
y_data = np.array(y_data, dtype = 'float32')
```

```python
from sklearn.preprocessing import normalize
normalize(data, norm = 'l2', axis =1)
```

```python
X_train,X_test,y_train,y_test = train_test_split(data,y_data,test_size = 0.1,random_state = 2)
```

```python
y_test
```

```python
# build model
from time import time
from sklearn.linear_model import LogisticRegression

it_times = []
acc = []

for tms in range(30):
    t = time()
    lr = LogisticRegression(max_iter= tms)
    lr.fit(X_train, y_train)
    t = time()-t 
    acc.append(lr.score(X_test,y_test))
    it_times.append(tms)
    print("Iterate times: ",tms)
    print("Time cost :",t)
    print("Accuracy :",acc,"\n")
    
plt.plot(it_times,acc)
```
