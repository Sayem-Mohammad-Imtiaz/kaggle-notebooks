```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv("../input/order-brushing-dataset-shopee-code-league-week-1/order_brush_order.csv")
```

```python
df['event_time'] = pd.to_datetime(df['event_time'])
```

```python
print(df.dtypes)
```

```python
lili = df.groupby(['shopid','userid','event_time'])['orderid'].count().reset_index().sort_values(by = ['shopid','event_time','userid'], ascending=[True , False, True])
lili.head()
```

```python
lili['datehour'] = lili['event_time'].dt.hour
lili.head()
```

```python
shopIdlist = df['shopid'].sort_values().unique()
print(shopIdlist)
```

```python
final_result = pd.DataFrame(columns= ["shopid","userid"])
final_result
```

```python
final_result1 = pd.DataFrame(columns= ["shopid","userid"])
final_result1
```

```python
dic = {}
date = 0

# Looping through the shopID unique list
for shopid in range(len(shopIdlist)):
    
    # Choosing the dataframe that have matches the the shopID list
    df_shop = lili.loc[lili['shopid'] == shopIdlist[shopid]]
    
    # Generate the list of date unique
    list_date = df_shop['event_time'].unique()
    
    # Skip the element if the there is just below 0 buyers in that shop within 1 hour
#     if len(df_shop) == 0:
#         new_row = {'shopid': shopIdlist[shopid], 'userid': 0}
#         final_result = final_result.append(new_row,  ignore_index=True)
#         continue
    
    for i in range(len(list_date)):

        # Take columns from the specific event_time to the last of it
        df_shop_date = df_shop.loc[df_shop['event_time'] <= list_date[i]]

        for index, row in df_shop_date.iterrows():
            
            
            # If date not have, assign to the current event_time
            if date == 0:
                dic[row['userid']] = 1
                date = row['event_time'].to_pydatetime()
            else:
                next_date = row['event_time'].to_pydatetime()
                
                second = abs((date - next_date).total_seconds())
                

                if second < 3600:
                    if row['userid'] not in dic:
                        dic[row['userid']] = 0
                    else:
                        dic[row['userid']] += 1
                else:
                    # Reset the sum of seconds
                    # hour = 0
                    
                    # Check if within 1 hour, if there just 2 order below
                    if len(dic) <= 0:
                        dic.clear()
                        date = 0
                        break
                        
                    # Get the number of unique user within 1 hour in that shop
                    uniqueUser = len(dic)
                    
                    # Get the number of order within 1 hour in that shop
                    sum_order = sum(dic.values())
                    
                    # Get the maxium number of that user's order within 1 hour in that shop
                    max_order = max(dic.values())
                    
                    # Calculate the concentrate_rate
                    concentrate_rate = sum_order / uniqueUser
                    
                    if concentrate_rate < 3:
                        new_row = {'shopid': shopIdlist[shopid], 'userid': 0}
                        final_result = final_result.append(new_row,  ignore_index=True)
                    else:
                        
                        for k, v in dic.items():

                            if v == max_order:

                                new_row = {'shopid': shopIdlist[shopid], 'userid': k}
                                final_result = final_result.append(new_row,  ignore_index=True)
                       
                    dic.clear()
                    date = 0
                    break
```
