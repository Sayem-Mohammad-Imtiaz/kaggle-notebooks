```python
plt.figure(figsize=(12,7))
plt.style.use("fivethirtyeight")
sns.lineplot(x='Age',y='Polyuria',data=data,hue='Gender')
```

```python
plt.figure(figsize=(12,7))
sns.histplot(data=data, x="Polydipsia",y='Age', hue="Gender", multiple="stack")
```

```python
f, axs = plt.subplots(1, 2, figsize=(12, 7), gridspec_kw=dict(width_ratios=[4, 3]))
sns.scatterplot(data=data, x="Age", y="weakness", hue="Gender", ax=axs[0])
sns.histplot(data=data, x="Polyphagia",y="Age" ,hue="Gender", shrink=.8, alpha=.8, legend=False, ax=axs[1])
f.tight_layout()
```

```python
plt.figure(figsize=(12,7))
plt.style.use("fivethirtyeight")
sns.lineplot(x='Age',y='class',data=data,hue='Gender')
```

```python
fig,axis=plt.subplots(2,2 , figsize=(20,10))
plt.style.use("fivethirtyeight")
sns.lineplot(x='Age',y='visual blurring',data=data,hue='Gender',ax=axis[0,0])
sns.lineplot(x='Age',y='Itching',data=data,hue='Gender',ax=axis[0,1])
sns.lineplot(x='Age',y='Irritability',data=data,hue='Gender',ax=axis[1,0])
sns.lineplot(x='Age',y='delayed healing',data=data,hue='Gender',ax=axis[1,1])
```
