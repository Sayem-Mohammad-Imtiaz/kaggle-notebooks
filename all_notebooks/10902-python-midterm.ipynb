## 相關說明
大樂透是一種樂透型遊戲。您必須從01~49中任選6個號碼進行投注。開獎時，開獎單位將隨機開出六個號碼加一個特別號，這一組號碼就是該期大樂透的中獎號碼，也稱為「獎號」。您的六個選號中，如果有三個以上（含三個號碼）對中當期開出之六個號碼（特別號只適用於貳獎、肆獎、陸獎和柒獎），即為中獎，並可依規定兌領獎金。

各獎項的中獎方式如下表：

![](https://i.imgur.com/sttoyUH.png)

```python
# 第三題檢查程式 --- 下列程式請勿修改
# !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
'''
================================================================================
第三題檢查程式 --- 只列出錯誤的資訊
測試所可能的位置如下所示 位置 ('a', 1)  -- ('h', 8) 
棋子 象 位置 ('a', 1)  
回傳結果 [('b', 2), ('c', 3), ('d', 4), ('e', 5), ('f', 6), ('g', 7), ('h', 8)]  
正確答案 [('b', 2), ('c', 3), ('d', 4), ('e', 5), ('f', 6), ('g', 7), ('h', 8)] 
檢查結果 正確
...
================================================================================
'''
def take_stepans(step):
    xi = step[step.str.contains("\('x', 'x'\)")].index
    step = step.drop(xi)
    pos = stepstr_totuple(step[0])
    steps = step[1:len(step)]
    ans_tuple = []
    for st in steps:
        ans_tuple.append(stepstr_totuple(st))
    return pos, ans_tuple  
    
def stepstr_totuple(pstr):
    pos = pstr[1:len(pstr)-1].replace("'", "").replace(" ","").split(',')
    return (pos[0], (int)(pos[1]))

# check answer
if piece == '馬':
    df = pd.read_csv("../input/chess-movestep/horse.csv", header=None)
elif piece == '象':
    df = pd.read_csv('../input/chess-movestep/elephant.csv', header=None)
chk_ok = 0
chk_false = 0
for i in range(len(df)):
    pos, ans = take_stepans(df.iloc[i].dropna())
    #print(pos, ans)
    res = NextStep(pos)
    chk_res = "錯誤"
    if sorted(res) == sorted(ans):
        chk_res = "正確"
        chk_ok += 1
    else:    
        chk_false += 1
        print(f"棋子 {piece} 位置 {pos}  \n回傳結果 {sorted(res)}  \n正確答案 {sorted(ans)} \n檢查結果 {chk_res}  ")
        print("================================================================================")
print("============================")
print(f"正確數目 {chk_ok}")
print(f"錯誤數目 {chk_false}") 
print("============================")
#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!    
```
