```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("train_age_dataset.csv")
df.info()
```

```python
df['age_group'].unique()
```

```python
df = df.iloc[:, 3:]
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
corr = df.corr()
corr_hm = sns.heatmap(corr, linecolor='white', cbar=True, linewidths=0.01)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=5)
X_pca = pca.fit_transform(X)
X_pca_df = pd.DataFrame(data=X_pca, columns=["PC1", "PC2", "PC3", "PC4", "PC5"])
X_pca_df.head()
```

```python
print(f'Explained variation per principal component: {pca.explained_variance_ratio_}')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_pca_df, y, test_size=0.33, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knearest = KNeighborsClassifier()
knearest.fit(X_train, y_train)
y_pred = knearest.predict(X_test)

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.neighbors import NearestNeighbors
nearest = KNeighborsClassifier()
nearest.fit(X_train, y_train)
y_pred = nearest.predict(X_test)

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.naive_bayes import BernoulliNB
clf = BernoulliNB()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
from sklearn.model_selection import train_test_split
X_train_n, X_test_n, y_train_n, y_test_n = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(X_train_n, y_train_n)
y_pred_n = clf.predict(X_test_n)

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test_n, y_pred_n)

from sklearn.metrics import accuracy_score
accuracy_score(y_test_n, y_pred_n)
```
