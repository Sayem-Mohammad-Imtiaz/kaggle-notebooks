# **Visualising our data**

Finding the total number of songs per Genre represented by top 50

```python
plt.figure(figsize=(6,7))
a=sns.countplot(data.Genre)
a.set(xlabel='Genre',
       ylabel='Total number of songs',
       title='Number of Songs represented by Genre');
plt.xticks(rotation=90)
plt.show()
```

Visualising beats per minute for different genre's

```python
plt.figure(figsize=(6,6))
sns.barplot(x=data['Genre'], y=data['Beats.Per.Minute'], palette='ocean')
plt.xticks(rotation=90)
plt.show()
```

Visualising Liveness of songs produced by different Artists 

```python
plt.figure(figsize=(6,6))
ax=plt.axes()
ax.scatter(data['Artist.Name'], data['Liveness'])
ax.set(xlabel='Artist Name',
       ylabel='Liveness',
       title='Liveness in Artits songs');
plt.xticks(rotation=90)
plt.show()
```

Visualisation for better understanding the features of the dataset

```python
sns.pairplot(data, plot_kws=dict(alpha=0.1, edgecolor='black'))
```

Checking if artists songs have varying popularity

```python
plt.figure(figsize=(6,6))
bx=plt.axes()
bx.scatter(data['Artist.Name'],data['Popularity'])
bx.set(xlabel='Artist name',
       ylabel='Popularity',
       title='Popularity by Artist');
plt.xticks(rotation=90)
plt.show()
```
