```python
flyers = pd.read_csv("/kaggle/input/d/peterpoydenko/goals-againstfor-flyers-12-march-april/flyers-stats.csv")
flyers
```

```python
flyers = flyers[:11]
flyers
```

```python
gf11 = [sum(flyers.p1f), sum(flyers.p2f), sum(flyers.p3f), sum(flyers.p1f) + sum(flyers.p2f) + sum(flyers.p3f)]
ga11 = [sum(flyers.p1a), sum(flyers.p2a), sum(flyers.p3a), sum(flyers.p1a) + sum(flyers.p2a) + sum(flyers.p3a)]
index = ['Period 1', 'Period 2', 'Period 3',
         'Total']
periodGoals = pd.DataFrame({'For': gf11,
                   'Against': ga11}, index = index)
ax = periodGoals.plot.bar(rot = 0, color=['orange','grey'])
ax.set_title('Flyers Goals For/Against Past Eleven Games')
ax.set_ylabel('Goals')
```

```python
#x controls the latest game included
#y controls the oldest game included

def gfgappbar(x, y):
    length = len(flyers)

    if (length - x - y > 0):
        gf11 = [sum(flyers.p1f[x:length-y]), sum(flyers.p2f[x:length-y]), sum(flyers.p3f[x:length-y])]
        ga11 = [sum(flyers.p1a[x:length-y]), sum(flyers.p2a[x:length-y]), sum(flyers.p3a[x:length-y])]
        index = ['Period 1', 'Period 2', 'Period 3']
        periodGoals = pd.DataFrame({'For': gf11,
                           'Against': ga11}, index = index)
        ax2 = periodGoals.plot.bar(rot = 0, color=['orange','grey'])
        ax2.set_title('Flyers Goals For/Against for '+ str(length-x-y) + ' Games of 2020/21 season')
        ax2.set_ylabel('Goals')
    else:
        print("bad values for x and y")
```

```python
#let's see how their last 5 games look
gfgappbar(0, len(flyers)-5)
```

Clearly, most of the flyers are late to their games so they lose the first period.
