# Predict the salary based on the number of years of experience

This is my first ML project and is beginner friendly. I have used Linear Regression to predict the salary of employees using an existing data set. I have trained the model using 80% of training data and tested the model using the rest. Plotted the actual data and predicted data 

### Import the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
sns.set()
```

### Read the input data set

```python
raw_data=pd.read_csv('../input/salary/Salary.csv')
raw_data.head()
```

### Understand the data

```python
raw_data.describe(include='all')
```

### See the data in scatter plot 

```python
sns.scatterplot(data=raw_data, y="Salary", x="YearsExperience")
#shows as years increase, salary also increases
```

### Make input and target

```python
x= raw_data['YearsExperience']
x_matrix=x.values.reshape(-1,1)   #need to reshape to avoid error in regression step. This wont happen if u have many input variables
y= raw_data['Salary']
```

### Split data into Train and Test

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_matrix,y,test_size=0.2)
```

### Train the model using Training data

```python

lm=LinearRegression()
lm.fit(x_train,y_train)
```

### Coef and Intercept

```python
lm.coef_
```

```python
#values from 0 to 1
#0 model explain None of the variability
#1 model explain Entire of the variability
lm.score(x_matrix,y)
```

```python
lm.intercept_  #B0 value
```

### Predict using Test data

```python
predictions = lm.predict(x_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.regplot(y_test,predictions)
```

```python

#AttributeError: 'Series' object has no attribute 'flatten' --to avoid this error in the next step
y_test.shape
y_test_matrix = y_test.values.reshape(-1,1)
y_test_matrix
```
