```python
df = pd.read_csv("../input/vnjobdes/jd.csv")
```

```python
cols = ['Địa điểm tuyển dụng ',
       'Tỉnh thành tuyển dụng ', 'Chức vụ ', 'Mức lương ',
       'Hình thức làm việc ', 'Ngành nghề ', 'Lĩnh vực ', 'Kinh nghiệm ',
       'Mô tả công việc ', 'Y/c công việc ', 'Quyền lợi được hưởng ']
len(df)
```

```python
df['Id tin '] = df['Id tin '].fillna('.')
for col in cols:
    df[col] = df[col].fillna('')
```

```python

def format_file(df):
    idx = 0
    index = idx
    while True:
        if (df['Id tin '][idx] != '.'):
            index = idx+1
            while(df['Id tin '][index]=='.'):
                for col in cols:
                    df[col][idx] = str(df[col][idx]) + str(df[col][index])
#                 df['Y/c công việc '][idx] = df['Y/c công việc '][idx] + df['Y/c công việc '][index]
#                 df['Mô tả công việc '][idx] = df['Mô tả công việc '][idx] + df['Mô tả công việc '][index]
#                 df['Ngành nghề '][idx] = df['Ngành nghề '][idx] + df['Ngành nghề '][index]
                index = index + 1
                if(index == 51270):
                    break;
            idx = index
            if (idx == 51270):
                break;
        else:
            idx = idx+1
    df = df.drop(df.loc[df['Id tin ']=='.'].index)
    return df

new_df = format_file(df)
            
```
