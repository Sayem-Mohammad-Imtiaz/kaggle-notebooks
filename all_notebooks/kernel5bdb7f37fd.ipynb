```python
df_train = pd.read_csv('/kaggle/input/heartbeat/mitbih_train.csv')
```

```python
df_train.shape
```

```python
df_test = pd.read_csv('/kaggle/input/heartbeat/mitbih_test.csv')
```

```python
df_train1 = np.concatenate((np.array(df_train.values),np.array(df_test.values)),axis=0)
df_train1.shape
```

```python
ytrain = df_train1[:,-1]
ytrain[ytrain>0] = 1
df_train1[:,:-1].shape, ytrain.shape
#df_train1 = np.concatenate((df_train1[:,:-1], ytrain), axis=0)
#df_train1
data = []
for i in range(df_train.shape[0]):
    data.append((list(df_train1[i])+[ytrain[i]]))
data = np.array(data)
data.shape
```

```python
np.sum(1-data[:,-1]), np.sum(data[:,-1])
```

```python
# x1 = np.array(df_train1[ytrain==1])
# print(x1.shape)
# x2 = np.array(df_train1[ytrain==2])
# print(x2.shape)
# x4 = np.array(df_train1[ytrain==4])
# print(x4.shape)
x0 = data[:72470]
x0 = np.array(x0)[np.random.random_integers(0,72469,14917)]
data = np.concatenate((x0,data[72470:]),axis=0)
data.shape
```

```python
# x0s = np.array(x0)[np.random.random_integers(0,72469,12000)]
# print(x0s.shape)
np.sum(1-data[:,-1]), np.sum(data[:,-1])
```

```python
x_train_final = np.copy(data)
np.random.shuffle(x_train_final)
x_train_final.shape
```

```python
traincount = 100
xtrain = x_train_final[:traincount,:-1]
ytrain = x_train_final[:traincount,-1]
xtest = x_train_final[traincount:,:-1]
ytest = x_train_final[traincount:,-1]
(xtrain.shape,ytrain.shape,xtest.shape,ytest.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
mdl = LogisticRegression(solver='newton-cg',multi_class='auto')
```

```python
mdl.fit(xtrain,ytrain)
```

```python
mdl.score(xtrain,ytrain)
```

```python
ypred = mdl.predict(xtest)
mdl.score(xtest,ytest)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(ytest,ypred)
```
