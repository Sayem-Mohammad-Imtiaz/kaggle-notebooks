```python
df=pd.read_csv("../input/restaurant-business-rankings-2020/Top250.csv")

x = df.Sales.values.reshape(-1,1)
y = df.Units.values.reshape(-1,1)
```

```python
plt.scatter(x,y)
plt.xlabel("sales")
plt.ylabel("Unites")

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x,y)
y_head=lr.predict(x)
plt.plot(x,y_head,color="red")
```

```python
print(lr.predict(np.array(10000).reshape(-1,1)))
```

```python
print("r^2 score: ", lr.score(x,y))
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg = DecisionTreeRegressor()
tree_reg.fit(x,y)
print(tree_reg.predict(np.array(10000).reshape(-1,1)))

x_=np.arange(min(x),max(x),1).reshape(-1,1)
y_head2=tree_reg.predict(x_)
plt.scatter(x,y,color="black")
plt.plot(x_,y_head2,color="yellow")
plt.xlabel("Sales")
plt.ylabel("Unites")
plt.show()
```
