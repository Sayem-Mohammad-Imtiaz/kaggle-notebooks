Let's explore this dataset and make some visualizations.

### *First of all necesary imports:*

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### *Read the data from csv*

```python
df = pd.read_csv('../input/911.csv')
```

#### *Info and head of the our Dataframe*

```python
df.info()
```

```python
df.head()
```

#### *At the moment Dataframe has data from Dec 2015 to Jan 2017, so lets convert timeStamp to datetime format and split Year to new column and creats new dataframe for 2016 year only*

```python
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
df['year'] = df['timeStamp'].apply(lambda x: x.year)
df2016 = df[df['year'] == 2016].copy()
```

#### *Let's check head and tail of our new dataframe to confirm it's 2016 only*

```python
df2016.head()
```

```python
df2016.tail()
```

### *Now we'll create few new features based on 'timeSTamp' to use later on for visualisation*

```python
df2016['month'] = df2016['timeStamp'].apply(lambda x: x.month)
df2016['hour'] = df2016['timeStamp'].apply(lambda x: x.hour)
df2016['day of week'] = df2016['timeStamp'].apply(lambda x: x.strftime('%a'))
df2016['date'] = df2016['timeStamp'].apply(lambda x: x.date())
df2016.head()
```

#### *Also let's split calls by types making a new feature from 'title'*

```python
 df2016['type'] = df2016['title'].apply(lambda x: x.split(':')[0])
```

#### *Now let's count calls by type and plot them*

```python
df2016['type'].value_counts()
```

```python
sns.countplot(df2016['type'])
```

### *Let's see distribution of call types by hours, weekdays and months*

```python
plt.figure(figsize=(11,6))
sns.countplot(df2016['hour'], hue=df2016['type'])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize=(8,6))
sns.countplot(df2016['day of week'], hue=df2016['type'])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
plt.figure(figsize=(11,6))
sns.countplot(df2016['month'], hue=df2016['type'])
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

### *Let's see number of calls by dates*

```python
df2016.groupby('date').count()['e'].plot(figsize=(10,5))
```

#### *And distribution by type*

```python
plt.figure(figsize=(11,8))
df2016[df2016['type'] == 'EMS'].groupby('date').count()['e'].plot()
df2016[df2016['type'] == 'Fire'].groupby('date').count()['e'].plot()
df2016[df2016['type'] == 'Traffic'].groupby('date').count()['e'].plot()
```

### *Now let's check what Township has most calls and plot top 10 of them*

```python
df2016.groupby('twp').count().sort_values('e', ascending=False)[0:10]
```

```python
twp = df2016.groupby('twp').count().sort_values('e', ascending=False)[0:10]
plt.figure(figsize=(11,8))
sns.barplot(x='twp', y='e', data=twp.reset_index())
plt.xticks(rotation=45)
```

### *Now lets create and plot Weekday|Hour matrix as heatmap*

```python
dayHour= df2016.groupby(by=['day of week', 'hour']).count()['e'].unstack()
dayHour
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(dayHour, cmap='YlGnBu')
```

#### *And the same for Weekday|Month*

```python
dayMonth= df2016.groupby(by=['day of week', 'month']).count()['e'].unstack()
dayMonth
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(dayMonth, cmap='viridis')
```
