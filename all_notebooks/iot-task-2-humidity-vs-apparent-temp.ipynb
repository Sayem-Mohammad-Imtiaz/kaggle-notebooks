```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
train = data.drop(['Temperature (C)', 'Apparent Temperature (C)', 'Formatted Date', 'Summary', 'Precip Type', 'Loud Cover', 'Daily Summary'], axis=1)
test = data['Apparent Temperature (C)']
```

```python
X_train, X_test, y_train, y_test = train_test_split (train, test, test_size=0.3, random_state=2)
```

```python
regrx= LinearRegression()
```

```python
regrx.fit(X_train, y_train)
```

```python
pred = regrx.predict(X_test)
```

```python
pred
```

```python
regrx.score(X_test, y_test)
```
