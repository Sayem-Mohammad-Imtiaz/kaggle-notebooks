Keskmise inimese(60kg) päevane kalorivajadus on 2100 (60x35) kcal

Eeldame, et inimene sööb 3x päevas ja iga toidukorra olulisus on 33%. Ehk siis meil on 700kcal, mis hommikul süüa.
proteiin ja süsivesikud on mõlemad 4kcal 1g kohta. rasvad on 9kcal 1g kohta
Süsivesikuid peaks selles olema 55-65%, rasvu 20-25% ja valke 10-15%. Samuti on vaja päevas saada 30g kiudaineid, ehk siis ühe toidukorraga 10g.
Ühe toidukorraga peaks saama 22g valke, 17,5g rasvu ja 105g süsivesikuid.

Uurime kas McDonaldsis oleks võimalik süüa tervislikus vahekorras olev hommikusöök? (mitte võttes arvesse valgu, süsivesikute ja rasvade kvaliteeti)
