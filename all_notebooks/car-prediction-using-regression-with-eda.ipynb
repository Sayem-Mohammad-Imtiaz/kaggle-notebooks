```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.formula.api as smf
```

```python
vif = pd.DataFrame()
vif['Features'] = x.columns
vif
```

```python
vif['VIF'] = [variance_inflation_factor(x_scaled, i) for i in range(x_scaled.shape[1])]
```

```python
vif
# we remove the features whose VIF > 5 i.e multicollinearity is present
# in this case there's no multicollinearity
```

```python
from sklearn.linear_model import Ridge, RidgeCV, Lasso, LassoCV, ElasticNet
```

```python
alphas = np.random.uniform(low=0, high=10, size=(50,))
ridgecv = RidgeCV(alphas = alphas,cv=10,normalize = True)
ridgecv.fit(xtrain, ytrain)
```

```python
ridgecv.alpha_
```

```python
ridgeReg = Ridge(alpha=ridgecv.alpha_, normalize=True)

ridgeReg.fit(xtrain,ytrain)

pred = ridgeReg.predict(xtest)
```

```python
ridgeReg.score(xtest, ytest)
```

```python
lasscv = LassoCV(alphas = alphas,cv =10, max_iter = 100000, normalize = True)
lasscv.fit(xtrain, ytrain)
```

```python
alpha = lasscv.alpha_
alpha
```

```python
lasso_reg = Lasso(alpha)
lasso_reg.fit(xtrain, ytrain)
```
