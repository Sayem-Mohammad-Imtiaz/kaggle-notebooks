# Ramen ratings



```python
# import the data
ramen_data = pd.read_csv('../input/ramen-ratings/ramen-ratings.csv')
# ramen_data.head()

# i just whant to have the style, the country, and the start
columns_to_keep = ['Style', 'Country', 'Stars']
ramen_data = ramen_data[columns_to_keep]

ramen_data.head()
```

```python
# choose the target and feature and deal with categorical data
Style_Columns = pd.get_dummies(ramen_data.Style, prefix = 'Style')
Country_Columns = pd.get_dummies(ramen_data.Country, prefix = 'Country')



y = ramen_data['Stars']

# deal with unrated value
y = y.replace(['Unrated'], '2' )
```

```python
X = pd.concat([Style_Columns, Country_Columns], axis = 1)
```

```python
# split the data into a test and val data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
# import the random forest regressor model that will be used
from sklearn.ensemble import RandomForestRegressor

ramen_model = RandomForestRegressor()
ramen_model.fit(X_train, y_train)
```

```python
# see how much our model work
from sklearn.metrics import mean_absolute_error
prediction = ramen_model.predict(X_test)

mea = mean_absolute_error(y_test, prediction)
```

```python
print(mea)
```

```python
# create an csv file with prediction
output_data = pd.DataFrame(prediction)
output_data.rename(columns={0: 'Prediction'}, inplace=True)

output_data.to_csv('outputfile.csv', index = False)
```
