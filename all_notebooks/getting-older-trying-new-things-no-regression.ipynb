```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')
```

# Import the data

```python
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv('/kaggle/input/have-you-ever-eaten-these-foods/responses.csv')
```

```python
colors = sns.color_palette("tab10")
```

# Grouping age by range

```python
def classify(Age):
    if Age<=20: return 'Teen'
    if 20<Age and Age<=30 : return 'Young Adult'
    if 30<Age and Age<=40 : return 'Mid Adult'
    if 40<Age and Age<=50 : return 'Old Adult'
    if Age>50: return 'Elderly'

data['Gen'] = data['Age in years'].apply(classify)
```

# Forward Fill the NaN 

```python
data_aged = data.sort_values(by='Age in years', ascending=True)
data_aged = data_aged.fillna(method='ffill')
```

# Plot the distribution of Age in dataset <br>
Then we manage to group the age by range defined above

```python
plt.hist(data_aged[data_aged['Gen']=='Teen']['Age in years'], color=colors[0], label='Teen')
plt.hist(data_aged[data_aged['Gen']=='Young Adult']['Age in years'],  color=colors[1], label='Young Adult')
plt.hist(data_aged[data_aged['Gen']=='Mid Adult']['Age in years'],  color=colors[2], label='Middle Adult')
plt.hist(data_aged[data_aged['Gen']=='Old Adult']['Age in years'],  color=colors[4], label='Old Adult')
plt.hist(data_aged[data_aged['Gen']=='Elderly']['Age in years'], color=colors[5], label='Elderly')
plt.xlabel('Age')
plt.legend()
plt.show()
```

# Create Food summary grouped by generation.

```python
FOOD = data_aged.drop(['Age in years','Gender','Race','Home location'], axis=1)
summary = FOOD.groupby(['Gen']).agg('sum')
summary = summary.reindex(['Teen', 'Young Adult', 'Mid Adult', 'Old Adult', 'Elderly'])
summary['count'] = FOOD.groupby(['Gen']).agg('count').iloc[:,0]
summary
```

```python
summary_ratio = summary.apply(lambda x:x/x['count']*100, axis=1)
summary_ratio.drop('count', axis=1, inplace=True)
summary_ratio.style.background_gradient(cmap=sns.light_palette("seagreen", as_cmap=True))
```

We'll investigate column by column and see if the color get darker downward. <br>
