```python
import matplotlib.pyplot as plt
meals = df.head(110)
meals_cal = pd.DataFrame({'Item': meals.Item, 'Calories': meals.Calories})
meals_sort = meals_cal.sort_values('Calories', ascending=False)
meals_sort.plot.barh(x='Item', y='Calories', figsize= (10,45))
plt.show()
```

```python
breakfast = df.loc[df.Category == 'Breakfast']
beefpork = df.loc[df.Category == 'Beef & Pork']
chickfish = df.loc[df.Category == 'Chicken & Fish']
salad = df.loc[df.Category == 'Salads']
snack = df.loc[df.Category == 'Snacks & Sides']
dessert = df.loc[df.Category == 'Desserts'] 
baverage = df.loc[df.Category == 'Beverages']
coffeetea = df.loc[df.Category == 'Coffee & Tea']
smoothies = df.loc[df.Category == 'Smoothies & Shakes']
```

```python
import seaborn as sns
fig, axes = plt.subplots(3, 3, figsize=(15, 7), sharex=True)
sns.color_palette("tab10")
sns.distplot( breakfast["Calories"] , color='black', ax=axes[0, 0], label = "Breakfast")
sns.distplot( beefpork["Calories"] , color='yellow',ax=axes[0, 1], label = "Beef & Pork")
sns.distplot( chickfish["Calories"] , color='green',ax=axes[0, 2], label = "Chicken & Fish")
sns.distplot( salad["Calories"] , color='purple',ax=axes[1, 0], label = "Salads")
sns.distplot( snack["Calories"] , color='blue',ax=axes[1, 1], label = "Snacks & Sides")
sns.distplot( dessert["Calories"] ,  color='violet',ax=axes[1, 2], label = "Desserts")
sns.distplot( baverage["Calories"] ,  color='red',ax=axes[2, 0], label = "Beverages")
sns.distplot( coffeetea["Calories"] ,  color='magenta',ax=axes[2, 1], label = "Coffee & Tea")
sns.distplot( smoothies["Calories"] ,  color='brown',ax=axes[2, 2], label = "Smoothies & Shake")
fig.suptitle("Calories Distribution on Each Menu Category")
fig.legend()
plt.show()
```

```python
avg_cat = [round(breakfast['Calories'].mean(axis=0), 2), round(beefpork['Calories'].mean(axis=0), 2), round(chickfish['Calories'].mean(axis=0), 2),
          round(salad['Calories'].mean(axis=0), 2), round(snack['Calories'].mean(axis=0), 2), round(dessert['Calories'].mean(axis=0), 2),
          round(baverage['Calories'].mean(axis=0), 2), round(coffeetea['Calories'].mean(axis=0), 2), round(smoothies['Calories'].mean(axis=0), 2)]
index = ['Breakfast', 'Beef & Pork', 'Chicken & Fish', 'Salads', 'Snacks & Sides', 'Desserts', 'Beverages', 'Coffee & Tea', 'Smoothies & Shakes']
avg_calat= pd.DataFrame({'Avg Calories': avg_cat}, index=index)
ax = avg_calat.plot.bar(rot=0, color='green', figsize=(13,8), title='Average Calories in Each Menu Category (in Cal)', legend=True)
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
baverages = df.loc[df.Category == 'Beverages']
cal_baverages = pd.DataFrame({'Item': baverages.Item, 'Calories': baverages.Calories})
cal_baverages['Men'] = cal_baverages.Calories/2500
cal_baverages['Women'] = cal_baverages.Calories/2000
cal_baverages
```

```python
avg_baverages = baverages['Calories'].mean(axis=0)
print("Average calories of 'Beverages' category is", round(avg_baverages, 2))
avg_baverages = pd.to_numeric(avg_baverages) 
baverages_men = avg_baverages/2500
print("Averagely 'Beverages' contributes to men's calories intake for ", round(baverages_men,2))

baverages_women = avg_baverages/2000
print("Averagely 'Beverages' contributes to women's calories intake for ", round(baverages_women, 2))
```

```python
crispy = df[df['Item'].str.contains('Crispy Chicken')]
crispy_cal = pd.DataFrame({'Item': crispy.Item, 'Calories': crispy.Calories})
crispy_sort = crispy_cal.sort_values('Calories', ascending=False)
ax = crispy_sort.plot.barh(x='Item', y='Calories', figsize= (10,7), color = 'purple')
ax.set_title("Total Calories in Crispy Chicken Item Category (in Cal)")
```

```python
avg_crispy_cal = crispy.Calories.mean(axis=0)
print("Average calories on Crispy Chicken category is ", avg_crispy_cal, "Cal.")
```

```python
grilled = df[df['Item'].str.contains('Grilled Chicken')]
grilled_cal = pd.DataFrame({'Item': grilled.Item, 'Calories': grilled.Calories})
grilled_sort = grilled_cal.sort_values('Calories', ascending=False)
ax = grilled_sort.plot.barh(x='Item', y='Calories', figsize= (10,7), color = 'yellow')
ax.set_title("Total Calories in Grilled Chicken Item Category (in Cal)")
```

```python
avg_grilled_cal = round(grilled.Calories.mean(axis=0), 2)
print("Average calories on Grilled Chicken category is ", avg_grilled_cal, "Cal.")
```

```python
avg_cal = pd.DataFrame({'Categories':['Crispy Chicken', 'Grilled Chicken'], 'Avg Calories': [avg_crispy_cal, avg_grilled_cal]})
ax = avg_cal.plot.bar(x = 'Categories', y = 'Avg Calories', figsize=(7,5), color = ['purple' , 'yellow'])
ax.set_title("Average Calories on Crispy Chicken and Grilled Chicken Menu Categories (in Cal)")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
whites = df[df['Item'].str.contains('Egg White')]
whites_cal = pd.DataFrame({'Item': whites.Item, 'Calories': whites.Calories})
whites_sort = whites_cal.sort_values('Calories', ascending=False)
whites_sort.plot.barh(x='Item', y='Calories', figsize= (10,7), color = 'magenta')
```

```python
whole = df[df['Item'].str.contains('Egg')] 
whole = whole[~whole['Item'].str.contains('White')]
whole_cal = pd.DataFrame({'Item': whole.Item, 'Calories': whole.Calories})
whole_sort = whole_cal.sort_values('Calories', ascending=False)
whole_sort.plot.barh(x='Item', y='Calories', figsize= (10,7), color = 'gold')
```

```python
avg_whites_cal = whites.Calories.mean(axis=0)
print("Average calories on Egg Whites category is ", round(avg_whites_cal, 2), "Cal.")
avg_whole_cal = whole.Calories.mean(axis=0)
print("Average calories on Whole Egg category is ", round(avg_whole_cal, 2), "Cal.")
```

```python
avg_egg_cal = [avg_whites_cal, avg_whole_cal]
index = ['Egg Whites', 'Whole Eggs']
avg_egg_cal = pd.DataFrame({'Avg Cal': avg_egg_cal,}, index=index)
ax = avg_egg_cal.plot.bar(rot=0 , color = 'cyan' )
ax.set_title("Average Calories in Egg Whites and Whole Eggs Categories (in Cal)")
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```
