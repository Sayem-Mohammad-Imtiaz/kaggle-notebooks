```python
# will ignore "id", as it is not required.
X = df.drop(['id', 'Class'], axis=1)
y = df['Class']
```

```python
X.shape, y.shape
```

```python
#split into train and test sets
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2, shuffle=True, random_state=99)
```

```python
X_train.shape, X_valid.shape, y_train.shape, y_valid.shape
```

```python
# Make the data scaler
from sklearn.preprocessing import StandardScaler
```

```python
# scale the X data    
sc = StandardScaler()
sc.fit(X_train) # fit only on train data
```

```python
X_train_sc = sc.transform(X_train)
```

```python
type(X_train_sc)
```

Usually we do transform the X_train, but this will convert our data set to numpy.ndarray.

So instead of just doing transform, lets convert it into daat frame so that we can refer to it easily.

```python
X_train = pd.DataFrame(sc.transform(X_train), index=X_train.index, columns=X_train.columns)
X_valid = pd.DataFrame(sc.transform(X_valid), index=X_valid.index, columns=X_valid.columns)
```

```python
X_train.head()
```

Lets start with LogisticRegression, which is a base model.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score
```

```python
lr = LogisticRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
pred_lr = lr.predict(X_valid)
```

```python
acc_lr = accuracy_score(y_valid, pred_lr)
acc_lr
```

```python
f1_lr = f1_score(y_valid, pred_lr, pos_label=1)
f1_lr
```

Very impressive... base model Logistic Regression itself has provided an accuracy of 98% and f1 score as 0.98.
