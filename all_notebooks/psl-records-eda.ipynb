```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")
import plotly
import plotly.express as px
import plotly.io as pio
pio.templates
from IPython.display import Image
!ls ../input/images
```

```python
df_highestBattingAvg = pd.read_csv('../input/psl-batting-records-20162021/HighestBattingAvgInPSL.csv')
df_mostRunsByBatsman = pd.read_csv('../input/psl-batting-records-20162021/MostRunsByBatsmanInPSL.csv')
df_highestIndvScore = pd.read_csv('../input/psl-batting-records-20162021/MostRunsInInnsInPSL.csv')
df_most50s = pd.read_csv('../input/psl-batting-records-20162021/Most50RunsInPSL.csv')
df_highestBattingAvg = pd.read_csv('../input/psl-batting-records-20162021/HighestBattingAvgInPSL.csv')
df_highestBattingSR = pd.read_csv('../input/psl-batting-records-20162021/HighestStrikeRateInPSL.csv')
df_allhundreds = pd.read_csv('../input/psl-batting-records-20162021/ListofHundredsInPSL.csv')
```

**Lets Start with First Analysis**

**Q-1: Who has Overall Highest Batting average in PSL (2016-2021)**

```python
df_highestBattingAvg.head()
```

```python
Top10Averages = df_highestBattingAvg[['PlayerName','Avg','Mat','Inns','HS','Span']].head(10)
fig = px.bar(Top10Averages,y="Avg",x="PlayerName", color="PlayerName",color_discrete_sequence=px.colors.qualitative.Antique,hover_data=['Mat','Inns','HS','Span'],title="Best Averages By Batsmans in PSL")
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.show()
```

**Lets Move to Question 2 Now**

**Q-2: Who has the highest Strike Rate in PSL (2016-2021)**

```python
df_highestBattingSR.head()
```

```python
Top10SR = df_highestBattingSR[['PlayerName','SR','Mat','Inns','HS','Span']].head(10)
fig = px.bar(Top10SR,y="SR",x="PlayerName", color="PlayerName",color_discrete_sequence=px.colors.qualitative.Antique,hover_data=['Mat','Inns','HS','Span'],title="Best Strike Rate By Batsmans in PSL")
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.show()
```

**Q3: List of all batsman pull a century in PSL (2016-2021)**

```python
df_allhundreds.head(9)
```

```python
Top10Hundreds = df_allhundreds[['PlayerName','Runs','BallsFaced','4s','6s','SR','Opposition','Venue']].head(9)
fig = px.bar(Top10Hundreds,y="Runs",x="PlayerName", color="PlayerName",color_discrete_sequence=px.colors.qualitative.Antique,hover_data=['BallsFaced','4s','6s','SR','Opposition'],title="All Hundreds in PSL")
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.show()
```

**Q:3: Most half centuries by a batsman in PSL (2016-2021)**

```python
df_most50s.head(10)
```

```python
print(len(df_most50s))
```

```python
Top50plusScores = df_most50s[['PlayerName','50+','Avg','Runs','Balls Faced','4s','6s','SR','Span','Inns','HS','0s']].head(10)
fig = px.bar(Top50plusScores,y="50+",x="PlayerName", color="PlayerName",color_discrete_sequence=px.colors.qualitative.Antique,hover_data=['Inns','Avg','HS','SR','4s','6s','0s'],title="Most 50+ Score By Batsmans in PSL")
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.show()
```

**Q:4: Most runs scored by a batsman in PSL (2016-2021)**

```python
df_mostRunsByBatsman.head()
```

```python
MostRunsInPSL = df_mostRunsByBatsman[['PlayerName','Mat','Span','100s','50s','Avg','Runs','Balls Faced','4s','6s','SR','Inns','HS','0s']].head(10)
fig = px.bar(MostRunsInPSL,y="Runs",x="PlayerName", color="PlayerName",color_discrete_sequence=px.colors.qualitative.Antique,hover_data=['Inns','Avg','HS','SR','4s','6s','0s','Span'],title="Most Runs Score By a Batsmans in PSL")
fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False)
fig.show()
```
