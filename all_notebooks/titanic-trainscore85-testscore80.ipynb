```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.utils import resample
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler,MinMaxScaler,PolynomialFeatures
from sklearn.svm import SVC
import sklearn.metrics as metrs 
import os
from sklearn.ensemble import GradientBoostingClassifier
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
#-------------------------Read The data----------------
data=pd.read_csv('//kaggle//input//titanic//train_and_test2.csv')
print(data.columns)



```

```python
#-------------------data QC and visualization----------------
print(data.agg(['min','max']))
print('#-------------------------------')
    
data =data.drop(['Passengerid' ,'zero', 'zero.1','zero.2', 'zero.3', 'zero.4', 'zero.5', 'zero.6', 'zero.7','zero.8', 
                     'zero.9', 'zero.10', 'zero.11', 'zero.12', 'zero.13','zero.14', 'zero.15', 'zero.16', 'zero.17','zero.18'],axis=1)#----drop all 'zero'columns

print(data.agg(['min','max']))#-------- after drop columns

print('#-------------------------------')    
#----------------------check nan values---------------
print(data.isnull().sum())
print('#------------------------')

nanv=data[data.isna().any(axis=1)]
print(nanv)
data=data.drop(nanv.index)#------------------dron nan values
               
print('#------------------------')
print(data.isnull().sum())
```

```python
corr=(data.corr())
plt.figure(figsize=(10,10))
sea.heatmap(corr,annot=True)
```

```python
#----------------Model prepration----------
y=data['2urvived']
X=data.iloc[:,:-1]

#------------------Encoding---------------------------------

Xencoding=pd.get_dummies(X,columns=[ 'sibsp','Parch','Pclass','Embarked'])
print(Xencoding.columns)

#-----------------------Scale the data-----------------
Xscaling=StandardScaler().fit_transform(Xencoding)

#-----------------------Data spliting-------------------
Xtrain,Xtest,ytrain,ytest=train_test_split(Xscaling,y,random_state=42)

print(Xtrain.shape)
print(Xtest.shape)
print(ytrain.shape)
print(ytest.shape)
```

```python
#-----------------GBoosting----------------

gb=GradientBoostingClassifier(n_estimators=100,random_state=42).fit(Xtrain,ytrain)
print('TrainScore--------------',gb.score(Xtrain,ytrain))
print('TestScore--------------',gb.score(Xtest,ytest))
```

```python
#-------------------------Model QC and evaluation------------------
metrs.plot_confusion_matrix(gb,Xtest,ytest)
metrs.plot_roc_curve(gb,Xtest,ytest)

ypred=gb.predict(Xtest)
print('#------------------------')
print(metrs.classification_report(ytest, ypred))
```
