```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stat
```

<a href="https://www.kaggle.com/kirilspiridonov/freelancers-offers-on-fiverr">The dataset</a>

```python
fiverr = pd.read_csv('/kaggle/input/freelancers-offers-on-fiverr/fiverr.csv')
fiverr
```

Listing of available categories in the sample:

```python
i=1
for categ in fiverr['Category'].unique():
    print(str(i)+' - '+str(categ))
    i=i+1
```

Listing of available subcategories for each category in the sample:

```python
i,j=1,1
for categ in fiverr['Category'].unique():
    print(str(i)+' - '+str(categ)+' \n')
    i=i+1
    j=1
    for subcateg in fiverr[fiverr['Category']==categ]['Subcat'].unique():
        print('    '+str(j)+' - '+str(subcateg))
        j=j+1
    print('\n')    
```

```python
fiverr ['price_description'] = fiverr['price']
fiverr['price'] = fiverr[fiverr['price']!='']
```

```python
def replace_comma_point(row):
    s = str(row['price_description'])
    return s.replace(',','.')
fiverr['price'] = fiverr.apply(replace_comma_point,axis=1)
fiverr['price']
```

```python
def last_index_num(text):
    index = -1
    for charcter in text:
        if charcter.isdigit():
            index = text.index(charcter)
    return index

def del_text(row):
    s = str(row['price'])
    x = last_index_num(s)
    return s[0:x]
fiverr['price'] = fiverr.apply(del_text,axis=1)
fiverr['price']
```

```python
def price_to_float(row):
    s = str(row['price'])
    return float(s[s.index('€')+1:])

fiverr['price'] = fiverr.apply(price_to_float,axis=1)
fiverr['price']
```

```python
title_font_size = 24
label_font_size = 16
plt.rcParams["figure.figsize"] = (18,6)
sns.kdeplot(fiverr['price'])
plt.title('Overall distribuion of Price')
```
