Ahora vamos a definir nuestro Target y variables:

```python
X=vinos.iloc[:, :-1].values
y=vinos["quality"]
```

Separemos nuestros data-sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Como dice el dermatólogo, vamos al grano

```python
from sklearn.linear_model import LinearRegression
regresion = LinearRegression()
regresion.fit(X_train, y_train)
```

```python
y_pred = regresion.predict(X_test)
np.std(y_pred-y_test)
```

Gráficamente podemos ver cómo se comparan nuestros resultados:

```python
ejex=np.linspace(0,2*np.max(y_pred),len(y_pred))
plt.plot(ejex,y_pred,c="r", label="predicha")
plt.plot(ejex,y_test,c="g", label="correctas");plt.legend()
```

Ahora vamos a imprimir qué pesos se le asignan a 

```python
sk_theta = list(regresion.coef_)
resultados=pd.DataFrame({'Peso':sk_theta,'Columns':columnas[:-1]})
```

Ahora, los pesos que buscamos en orden (según valor absoluto) son:

```python
resultados.reindex(resultados.Peso.abs().sort_values(ascending = False).index)
```
