```python
import matplotlib.pyplot as plt
df=pd.read_csv('/kaggle/input/loan-predication/train_u6lujuX_CVtuZ9i (1).csv')
df.head(10)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df1=df.dropna()
df1.shape
```

```python


df2 = df1.replace(to_replace = ['Yes','No'],value = ['1','0'])
df2.head()
```

```python
df2 = df1.replace(to_replace = ['Y','N'],value = ['1','0'])
df2.head()
```

```python
X = df2[['ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term']]
y=df2.Loan_Status

```

```python
y=df2.Loan_Status
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

# RandomForestClassifier

```python
#RandomForestClassifier
from sklearn.ensemble import RandomForestClassifier
Classifier=RandomForestClassifier(n_estimators=20)
Classifier.fit(X_train,y_train)
```

```python
Classifier.score(X_test,y_test)
```

```python
y_predicted= Classifier.predict(X_test)
y_predicted
```

```python
plt.figure(figsize=(10,7))
feat_importances= pd.Series(Classifier.feature_importances_,index=X_train.columns)
feat_importances.nlargest(7).plot(kind='barh');
```

```python
import pickle
pickle_out=open('classifier.pkl','wb')
pickle.dump(Classifier,pickle_out)
pickle_out.close()
```

```python
Classifier.predict([[4000,6000,120,120]])
```

# DecisionTreeClassifier

```python
#DecisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
# define the model
TreeClassifier = DecisionTreeClassifier()
# fit the model
TreeClassifier.fit(X_train,y_train)
TreeClassifier.score(X_test,y_test)
#score = accuracy_score(y_test,y_pred)
#score
```

```python
y_pred = TreeClassifier.predict(X_test)
y_pred
```

```python
plt.figure(figsize=(10,7))
feat_importances= pd.Series(TreeClassifier.feature_importances_,index=X_train.columns)
feat_importances.nlargest(7).plot(kind='barh');
```

# XGBClassifier

```python
#XGBClassifier
from xgboost import XGBClassifier
# define the model
xgclassifier = XGBClassifier()
# fit the model
xgclassifier.fit(X_train,y_train)
xgclassifier.score(X_test,y_test)
#score = accuracy_score(y_test,y_pred)
#score
```

```python
y_pred = xgclassifier.predict(X_test)
y_pred
```

```python
plt.figure(figsize=(10,7))
feat_importances= pd.Series(xgclassifier.feature_importances_,index=X_train.columns)
feat_importances.nlargest(7).plot(kind='barh');
```

# KNeighborsClassifier

```python
#KNeighborsClassifier
from sklearn.neighbors import KNeighborsClassifier
# define the model
knClassifier = KNeighborsClassifier()
# fit the model
knClassifier.fit(X_train,y_train)
knClassifier.score(X_test,y_test)
#score = accuracy_score(y_test,y_pred)
#score
```
