```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] =(15,8)
plt.style.use('ggplot')
```

```python
data = pd.read_csv('/kaggle/input/1500-gameboy-advanced-games-information/GBA_Games.csv')
```

```python
data.head(10)
```

```python
RR = []
for R in data['Region Released']:
    if type(R) == float:
        continue
    dr = R.split(',')
    RR = RR + dr
```

```python
plt.title('ditsribution of regions')
sns.countplot(x=RR)
```

```python
PUB = []
for R in data['Publishers']:
    if type(R) == float:
        continue
    dr = R.split('/')
    PUB = PUB + dr

PUB = pd.Series(PUB).value_counts()[:10]
```

```python
plt.title('top 10 publisher distribution')
sns.barplot(x=PUB.index,y=PUB.values)
```
