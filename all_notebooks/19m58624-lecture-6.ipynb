**I chose to present the COVID-19 trend in Tunisia**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.set_printoptions(threshold=np.inf)

selected_country='Tunisia'
df = pd.read_csv('../input/novel-corona-virus-2019-dataset/covid_19_data.csv', header=0)

del df['SNo']
df = df[df['Country/Region']==selected_country]
df = df.groupby('ObservationDate').sum()
print(df)
```

# **COVID-19 trend in Tunisia**

```python
df1 = df.diff()
print(df1.plot())
```

#  **Interactive chart**

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_confirmed_object = go.Scatter(x=df1.index, y=df1['Confirmed'].values, name='Confirmed')
daily_deaths_object = go.Scatter(x=df1.index, y=df1['Deaths'].values, name='Deaths')
daily_recovered_object = go.Scatter(x=df1.index, y=df1['Recovered'].values, name='Recovered')
layout_object = go.Layout(title='Tunisia Daily Cases',xaxis=dict(title='Date'),yaxis=dict(title='Number of People'))
fig = go.Figure(data=[daily_confirmed_object,daily_deaths_object,daily_recovered_object],layout=layout_object)
iplot(fig)
```

# **Informative Table of Covid-19 Cases in Tunisia**

```python
df1= df1.fillna(0.)
styled_object = df1.style.background_gradient(cmap='gist_ncar').highlight_max('Confirmed').set_caption('Daily Summaries')
display(styled_object)
f = open('table_19M58624.html','w')
f.write(styled_object.render())
```

# Tunisia's Global Ranking

```python
df = pd.read_csv('../input/novel-corona-virus-2019-dataset/covid_19_data.csv', header=0)

latest = df[df['ObservationDate']=='06/12/2020']
latest = latest.groupby('Country/Region').sum()
latest = latest.sort_values(by='Confirmed',ascending=False).reset_index()
print('Global Ranking of Tunisia: ', latest[latest['Country/Region']==selected_country].index.values[0]+1)
```

# Discussion

The first confirmed Covid-19 case in Tunisia was reported on 2 March 2020, with the patient being a man in his 40s returning from Italy. By the 18th of April 2020, all 24 states of Tunisia were affected by the pandamic before the spread of the virus started to bubside gradually. As of June 12, Tunisia recorded a total of 1093 (1904 on the 13th) confirmed cases, 49 deaths and 995 recoveries. 17 of Tunisia's 24 states still house active cases till this date.

Based on the charts above it can be observed that Tunisia's confirmed cases peaked quite early (end of March). This can be attributed to the swift action by the Tunisian authorities to stem the spread of the virus as well as the commitment of the public to the necessary social distancing measures. Following an emergency government meeting, all schools, universities and other education facilities were to be closed starting from March 12th. Stricter measures soon followed: closure of all land borders and air connextions (16/03/2020), 18 hours curfew starting 6am (18/03/2020).

The continuous decrease in confirmed cases as reflected in the flattening of the infection curve, has allowed the gradual return of daily activities and the easing of the state of health emergency starting May 4th. All curfews and restrictions on movement between cities were lifted on the 8th of June. After 8 days of zero new cases, 6 tunisians tested positive of the virus on the 12th.

Despite its apparent success in manageing the covid-19 pandemic (121 in global ranking), Tunisia has suffered tremendous financial losses, especially since its economy relies heavily on tourism (14% of GDP). Thus, the government is activaly considering strategies and sanitary protocols to attract foreign visitors and unsure public health. 78% of tunisians are optimistic about the post covid economy according to a pole by Emrhod Consulting agency.
