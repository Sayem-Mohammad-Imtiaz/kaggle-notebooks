### Import Libraries

```python
import pandas as pd
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

### Read Data

```python
df=pd.read_csv('../input/fantasy-premier-league-21-full-player-dataset/fpl_2021_player_gw.csv')
#view data head
df.head()
```

```python
#Pivot table get gameweek data for each player and team combo
df_sum=pd.pivot_table(data=df,index=['name','team'],columns='game_week', values='total_points', aggfunc='sum')
df_sum.fillna(0, inplace=True)
df_sum=df_sum.reset_index()
df_sum.head()
```

```python
#cumulative sum to get total points in each GW
for i in range(2,39):
    df_sum[i]+=df_sum[i-1]
df_sum.head()
```

```python
#Rename column Headers
cols={}
for i in range(1,39):
    cols.update({i:f'GW {str(i)}'})
    
df_sum=df_sum.rename(columns=cols)
```

```python
# Add image link
#Link to image scraped from premierleague website
#Images of three relegated teams not found!!!
#Since none of the players from the relegated teams ever made into the top scorers list, its not required
img_dict={'ARS': 'https://resources.premierleague.com/premierleague/badges/25/t3.png',
 'AVL': 'https://resources.premierleague.com/premierleague/badges/25/t7.png',
 'BHA': 'https://resources.premierleague.com/premierleague/badges/25/t36.png',
 'BUR': 'https://resources.premierleague.com/premierleague/badges/25/t90.png',
 'CHE': 'https://resources.premierleague.com/premierleague/badges/25/t8.png',
 'CRY': 'https://resources.premierleague.com/premierleague/badges/25/t31.png',
 'EVE': 'https://resources.premierleague.com/premierleague/badges/25/t11.png',
 'LEE': 'https://resources.premierleague.com/premierleague/badges/25/t2.png',
 'LEI': 'https://resources.premierleague.com/premierleague/badges/25/t13.png',
 'LIV': 'https://resources.premierleague.com/premierleague/badges/25/t14.png',
 'MCI': 'https://resources.premierleague.com/premierleague/badges/25/t43.png',
 'MUN': 'https://resources.premierleague.com/premierleague/badges/25/t1.png',
 'NEW': 'https://resources.premierleague.com/premierleague/badges/25/t4.png',
 'SOU': 'https://resources.premierleague.com/premierleague/badges/25/t20.png',
 'TOT': 'https://resources.premierleague.com/premierleague/badges/25/t6.png',
 'WHU': 'https://resources.premierleague.com/premierleague/badges/25/t21.png',
 'WOL': 'https://resources.premierleague.com/premierleague/badges/25/t39.png'}

#Function which takes the team as argument and return image url
def team_finder(x):
    try:
        return img_dict[x]
    except:
        return ""
    
#New column for image url    
df_sum['image']=""
#add image url
df_sum.image=df_sum.team.apply(lambda x:team_finder(x))

#new columns order put image after name and team
cols=df_sum.columns
cols_order=list(cols[:2])           #name, team
cols_order.append(cols[-1])         #image
cols_order.extend(list(cols[2:-1])) #gameweeks

#rearrage columns
df=df_sum[cols_order]
df.head()
```

```python
#save dataframe
df.to_csv('fl_data.csv', index=None)
```
