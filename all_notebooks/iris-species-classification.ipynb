```python
df = pd.read_csv('../input/iris-flower-dataset/IRIS.csv', header = 0)
print(df.shape)
print(df.species.value_counts())
print(df.columns)
df.head()
```

```python
def update_labels(x):
    if x.lower() == 'iris-setosa':
        return 0
    elif x.lower() == 'iris-virginica':
        return 1
    else:
        return 2

df['species'] = df['species'].map(lambda x: update_labels(x))
print(df.species.value_counts())

y = np.array(df.species)
y = to_categorical(y)
print(y[:10])
```

```python
df = df[df.columns[:4]]
df.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df, y, random_state = 20, test_size = 0.2)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
input_features = df.shape[1]
input_features
```

```python
def define_model():
    model = Sequential()
    model.add(Input(shape=(input_features,)))
    model.add(layers.Dense(64, activation = 'relu'))
    model.add(layers.Dense(3, activation = 'softmax'))
    
    model.compile(loss='categorical_crossentropy', optimizer= 'adam', metrics = ['accuracy'])
    return model

model = define_model()
model.summary()
```

```python
print("Entrenamiento:")
history = model.fit(X_train, y_train, epochs = 10, verbose = 1, validation_data= (X_test, y_test), batch_size=8)

# Verificamos el modelo con datos con los que no ha aprendido.
print('\nEvaluaci√≥n en conjunto de test:')
val_score = model.evaluate(X_test, y_test)
# print('Accuracy de la RNA en el conjunto de test: {}, equiv. {}%'.format(val_score[1], round(val_score[1]*100, 2)))
print('Accuracy de la RNA en el conjunto de test: {}, equiv. {:.2%}%'.format(val_score[1],val_score[1]))
```
