```python
df = pd.read_csv("../input/gufhtugu-publications-dataset-challenge/GP Orders - 5.csv", sep=",", encoding="utf_8")
df.rename({'ï»¿Order Number': 'Order Number',"Order Date & Time":"Order Date","City":"City (Billing)"}, axis=1, inplace=True)
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Order Date']).dt.date
df['Day'] = df['Order Date'].dt.day
df['Month'] = df['Order Date'].dt.month
df['Quarter'] = df['Order Date'].dt.quarter
df['Week'] = df['Order Date'].dt.week
df['Year'] = df['Order Date'].dt.year
df['Hour'] = df['Order Date'].dt.hour
df['Day_name'] = df['Order Date'].dt.day_name()
df['Weekday'] = pd.to_datetime(df['Order Date']).apply(lambda x: x.weekday())
# is_holiday flag
df['Holiday'] = df.Day_name.apply(lambda x : "Week End" if x == "Sunday" or x == "Saturday" else "Week Day")
df.head()
```

```python
import pandas as pd

if isinstance(df, (pd.DatetimeIndex, pd.MultiIndex)):
    df = df.to_frame(index=False)

df = df.reset_index().drop('index', axis=1, errors='ignore')
df.columns = [str(c) for c in df.columns]  # update columns to strings in case they are numbers

chart_data = pd.concat([
    df['Date'],
    df['Order Number'],
], axis=1)
chart_data = chart_data.sort_values(['Date'])
chart_data = chart_data.groupby(['Date'])[['Order Number']].count()
chart_data = chart_data.dropna()
chart_data.index = pd.to_datetime(chart_data.index)
chart_data.head()
chart_data.plot(title ="Date Wise Number of Orders", figsize=(30, 10), legend=True, fontsize=12)
```
