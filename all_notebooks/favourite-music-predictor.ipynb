```python
music_data=pd.read_csv('/kaggle/input/favouritemusic/music.csv')
music_data
```

```python
#make predictions for what age and gender like certain music

from sklearn.tree import DecisionTreeClassifier

X=music_data.drop(columns=['genre']) #create our input and output
y=music_data['genre']

model=DecisionTreeClassifier() #create a model
model.fit(X,y) #train it
predictions=model.predict([[21,1],[22,0]]) #make predictions for 21 year old male and 22 year old female
predictions
```

```python
#accuracy, test train split

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X=music_data.drop(columns=['genre'])
y=music_data['genre']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

model=DecisionTreeClassifier()
model.fit(X_train,y_train)
predictions=model.predict(X_test)

score=accuracy_score(y_test, predictions)
score
#very small features.  the accuracy score ranges from .75 to 1
```
