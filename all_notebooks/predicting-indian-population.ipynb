**Evaluate Fuction**

```python
from sklearn import metrics

def evaluate_model(y_true, y_predict):
    mae = metrics.mean_absolute_error(y_true, y_predict)
    mse = metrics.mean_squared_error(y_true, y_predict)
    mape = np.mean((abs((y_true - y_predict) / y_true))) * 100
    rmse = np.sqrt(metrics.mean_squared_error(y_true, y_predict))
    
    print('Model Performance')
    print('MAE = {:0.4f}'.format(mae))
    print('MAPE = {:0.4f}%'.format(mape))
    print('MSE = {:0.4f}'.format(mse))
    print('RMSE = {:0.4f}'.format(rmse))
    
    return mae, mse, mape, rmse
```

**Split Data**

```python
from sklearn.linear_model import LinearRegression

#Split Data
X_train = np.array(df['Year'][:-10]).reshape(-1,1)
y_train = np.array(df['Population'][:-10]).reshape(-1,1)

X_test = np.array(df['Year'][-10:]).reshape(-1,1)
y_test = np.array(df['Population'][-10:]).reshape(-1,1)
```

**Build The Models**

```python
#Build Linear Model
model_linear = LinearRegression()
model_linear.fit(X_train,y_train)
predict_model_linear = model_linear.predict(X_test)
evaluate_model(y_test, predict_model_linear)
```

```python
from sklearn.linear_model import Ridge

model_ridge = Ridge(alpha=1.0)
model_ridge.fit(X_train,y_train)
predict_model_ridge = model_ridge.predict(X_test)
evaluate_model(y_test, predict_model_ridge)
```

```python
from sklearn.linear_model import Perceptron

model_perceptron = Perceptron()
model_perceptron.fit(X_train,y_train)
predict_model_perceptron = model_perceptron.predict(X_test)
evaluate_model(y_test, predict_model_perceptron)
```
