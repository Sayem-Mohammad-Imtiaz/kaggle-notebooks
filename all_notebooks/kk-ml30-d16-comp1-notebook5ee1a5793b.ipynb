```python
train = pd.read_csv("../input/kk30ml/train.csv", index_col = 0)
test  = pd.read_csv("../input/kk30ml/test.csv", index_col = 0)

train
```

```python
test.head()
```

```python
y = train['target']
X = train.drop(['target'], axis = 1)
X.head()
```

```python
#preparing the data 
#preprocessing for categorical data 
from sklearn.preprocessing import OrdinalEncoder
#identifying categorical data in training set

cat = X.dtypes == 'object' #returns True if Categorical variables 
cat_l  = list(cat[cat].index)

#performing ordinal encoding 

X_Tr = X.copy()
X_test = test.copy()

ordinal_encoder = OrdinalEncoder()
X_Tr[cat_l] = ordinal_encoder.fit_transform(X[cat_l])
X_test[cat_l] = ordinal_encoder.transform(X_test[cat_l])

X_Tr.head()
```

```python
from sklearn.model_selection import train_test_split 

#train and valid split 

X_train, X_valid, y_train, y_valid = train_test_split(X_Tr, y, random_state = 0)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

#define the model

model = RandomForestRegressor()

#fit the model

model.fit(X_train, y_train)
pred = model.predict(X_valid)
mse = mean_squared_error(y_valid, pred, squared = False)
print(mse)
```

```python
#finally generating predictions on test data 

pred_final = model.predict(X_test)

#creating a csv file to submit 

op = pd.DataFrame({'Id' : X_test.index, 'target': pred_final})
op.to_csv('submission.csv', index = False)
```
