```python

def mod(x):
    return pd.Series(index=['mode'],
                    data=[x.mode()])
def mea(x):
    return pd.Series(index=['mean','var', 'st. desv.','min', 'max', 'range'],
                    data=[x.mean(), x.var(), x.std(), x.min(), x.max(), x.max()-x.min()])

df = pd.read_csv(r"/kaggle/input/the-human-freedom-index/hfi_cc_2020.csv")
#quantitative vars
dfnumbers=df.select_dtypes(exclude=['object'])
#categorical vars
dfobjects=df.select_dtypes(include=['object'])
dfnumbers

#all df mode
df.apply(mod)
#mean, var, std, min, max and range
dfnumbers.apply(mea)
```
