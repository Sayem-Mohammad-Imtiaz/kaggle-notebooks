```python
data = pd.read_csv("/kaggle/input/gufhtugu-publications-dataset-challenge/GP Orders - 5.csv")
data.shape
```

```python
data.head()
```

```python
data.describe()
```

```python
#Change Columns Name
data.rename( columns = {'Order Number' : 'OrderNo', 'Order Status' : 'OrderStatus', 'Book Name' : 'BookName', 
                        'Order Date & Time' : 'OrderDate', 'Payment Method' : 'PayMethod', 'Total items' : 'TotalItems', 
                        'Total weight (grams)' : 'Weight(g)'}, inplace=True )
```

```python
data.columns
```

```python
#Import DateTime Library and
import datetime as dt
data['OrderDate'] = pd.DatetimeIndex(data["OrderDate"])
```

```python
#Extract the Year, Month(in Number & Name),
#Date (Days, Week) and Time from Date Column
data['Years'] = data["OrderDate"].dt.year
data['Months'] = data["OrderDate"].dt.month_name()
data['Months'] = data["OrderDate"].dt.month
data['Days'] = data["OrderDate"].dt.day
data['DaysName'] = data["OrderDate"].dt.day_name()
data['Weeks'] = data["OrderDate"].dt.week
data['Date'] = data["OrderDate"].dt.date
#Confirm Extracted Columns
data.head(5)
```

```python
data = data.dropna()
```

```python
#import Library File
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style('whitegrid')
plt.title("Order Status (Frequency) ")
sns.countplot(x="OrderStatus", data=data)
```

```python
#Change in Lower-Case
data['PayMethod'] = data['PayMethod'].str.lower()
```

```python
data.PayMethod.unique()
```

```python
#Replace for Unique Vlaues
data['PayMethod'] = data['PayMethod'].replace('cash on delivery (cod)', 'cash on delivery')
```

```python
data.PayMethod.unique()
```

```python
sns.set_style('whitegrid')
plt.title("Payment Method (Frequency) ")
sns.countplot(x="PayMethod", data=data)
```

```python
sns.set_style('whitegrid')
plt.subplots(figsize=(10,5))
plt.title("Order Status and Payment Method")
sns.barplot(x="PayMethod", y="OrderNo", hue="OrderStatus", data=data)
```

```python
sns.set_style('whitegrid')
plt.subplots(figsize=(10,5))
plt.title("Yearly Orders Status (Frequency) ")
sns.barplot(x="Years", y="OrderNo", hue="OrderStatus", data=data)
```
