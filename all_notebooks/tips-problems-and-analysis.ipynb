## Problem 1 and Problem 2 
- Which day how much tip we earn?



- Which time of a day how much tip we earn?

In the same time we can answer the second problem with this chart

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df,x="day",y="tip",hue="time");
```

Answer= The order of the days and time of days we earn the most tip;

1. Saturday / Dinner
2. Sunday / Dinner
3. Thursday / Lunch
4. Friday / Dinner

## Problem 3 
- How much is the difference between the effect of smoker / non-smoker to the tip?

Total values= 

```python
s=df.groupby(["day","smoker"])
s.sum()
```

Median values=

```python
s.mean()
```

* ####  Answer=

Thursday= Smokers pays more tip 

Friday= Smokers and Non-Smokers pays equal tip

Saturday= Non-smokers pays more tip

Sunday= Smokers pays more tip

<b><i>Formula = ( [Non-smokers total_bill/Smokers total_bill] * Smokers tip ) - Non-smokers tip</i> </b>
 

## Problem 4

- Which gender pays more tip?

```python
plt.figure(figsize=(12,10))
sns.scatterplot(data=df,x="total_bill",y="tip",hue="sex");
```

* Answer= Males pays more tip than Females generally. 

## Problem 5
- At what time of the day smokers pays tip more ?



```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df,x="smoker",y="tip",hue="time");
```

* Answer= Non-smokers much more pays tip on Dinners, but no difference on smokers.
