```python
stars_float = []
for i in df['Stars']:
    if i=='Unrated':
            stars_float.append(-1.0)
    else:
        stars_float.append(float(i))
```

```python
df['Stars'] = stars_float
```

```python
df.head()
```

```python
data = df[df['Stars']!=-1]
```

```python
data.shape,df.shape
```

```python
mean = {}
for i in data['Country'].unique():
    data_temp = data[data['Country']==i]
    mean[i] = np.mean(data_temp['Stars'])
    
```

```python
print(mean)
```

```python
mean_sorted  =  dict(sorted(mean.items(), key=lambda item: item[1]))
```

```python
plt.figure(figsize = (14,8))
plt.title('Average Star Ratings of each country')
plt.xlabel('Stars')
plt.barh(*zip(*mean_sorted.items()))
plt.show()
```
