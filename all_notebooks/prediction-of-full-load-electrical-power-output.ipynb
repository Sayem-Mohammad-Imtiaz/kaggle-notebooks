# Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Import Dataset

```python
df = pd.read_csv('../input/combined-cycle-power-plant/Combined Cycle Power Plant Dataset.csv')
```

# Attribute Information:

Features consist of hourly average ambient variables
- Temperature (AT) in the range 1.81°C and 37.11°C,
- Ambient Pressure (AP) in the range 992.89-1033.30 milibar,
- Relative Humidity (RH) in the range 25.56% to 100.16%
- Exhaust Vacuum (V) in teh range 25.36-81.56 cm Hg
- Net hourly electrical energy output (PE) 420.26-495.76 MW
The averages are taken from various sensors located around the plant that record the ambient variables every second. The variables are given without normalization.

# Define x and y

```python
x = df.drop(['PE'],axis=1).values
y = df['PE'].values
```

```python
x
```

```python
y
```

# Split the dataset in training set and test set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=0)
```

# Train the model on the training set 

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train,y_train)
```

# Predict the test set result

```python
reg.predict(x_test)
```

```python
y_pred = reg.predict([[8.34,40.77,1010.84,90.01]])
y_pred
```

# Check Predict Result with Linear regression formula y = mx + c

```python
#Calculate Coefficient
reg.coef_
```

```python
#Calculate Intercept
reg.intercept_
```

```python
#y = (m1*AT + m2*V + m3*AP + m4*RH) + c(coefficient)
y = (-1.9797936*8.34 + 40.77* -0.23300225 + 0.06812315*1010.84 + 90.01* -0.15839461) +448.53067140557835
y
```

# Evaluate The Model And Get Accuracy

```python
from sklearn.metrics import r2_score
y_pred = reg.predict(x_test)
score = r2_score(y_test,y_pred)
score
```

# Plot the results

```python
plt.figure(figsize=(15,10))
plt.scatter(y_test,y_pred)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
```
