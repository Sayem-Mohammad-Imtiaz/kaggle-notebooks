# Exploratory Data Analysis
## Time series of vibration data
Let's follow the tutorial https://machinelearningmastery.com/time-series-data-visualization-with-python/

```python
# Select a subset of the vibration records
time_serie = dataset.iloc[1:1000,:]
```

## Graphics of temporal evolution: all in one graph
The next plots show the time series of the 4 accelerometers placed in the gearbox

```python
# Import Python libraries for graphical representations
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = 18, 8
time_serie.plot()
plt.show()
```

#### Plotting only points

```python
time_serie.plot(style='.')
plt.show()
```

## Graphics of temporal evolution: separated graphs
The next plots show the time series of the 4 accelerometers a1, a2, a3, a4 from top-left to bottom-right

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

a1 = time_serie.iloc[:,0]
a2 = time_serie.iloc[:,1]
a3 = time_serie.iloc[:,2]
a4 = time_serie.iloc[:,3]

a1.plot(ax=axes[0,0])
a2.plot(ax=axes[0,1])
a3.plot(ax=axes[1,0])
a4.plot(ax=axes[1,1])
```

## Histogram of the four vibration signals
Each plot shows the histogram for each of the 4 accelerations a1, a2, a3, a4

```python
time_serie.hist(bins=100)
plt.show()
```

## Plots of probability density functions of vibration signals
KDE Plot described as *Kernel Density Estimate* is used for visualizing the Probability Density of a continuous variable
### Based on 1000 points

```python
time_serie.plot(kind='kde')
plt.xlim(-20, 20)
plt.show()
```

### Based on the full dataset (100.000 points per signal)

```python
# dataset is the variable that host the full record for all 4 signals
## It takes a while to depict them...
dataset.plot(kind='kde')
plt.xlim(-20, 20)
plt.show()
```

## Autocorrelation for a1 vibration signal
y(t+1) vs. y(t) values

```python
from pandas.plotting import lag_plot

lag_plot(time_serie.iloc[:,0])
plt.show()
```

And as a function of the lag value: y(t+lag) vs. y(t) values

```python
from pandas.plotting import autocorrelation_plot

autocorrelation_plot(time_serie.iloc[:,0])
plt.show()
```
