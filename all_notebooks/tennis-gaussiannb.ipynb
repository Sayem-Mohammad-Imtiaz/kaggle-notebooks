Map values

```python
#map
outlook1={'Sunny':1, 'Overcast':2, 'Rain':3}
df.outlook=df.outlook.map(outlook1)

temp1={'Hot':1, 'Mild':2, 'Cool':3}
df.temp=df.temp.map(temp1)

humid1={'Normal':1, 'High':2}
df.humidity=df.humidity.map(humid1)

wind1={'Weak':1, 'Strong':0}
df.wind=df.wind.map(wind1)

play1={'Yes':1, 'No':0}
df.play=df.play.map(play1)

df
```

Heatmap

```python
#create a heatmap to correlate promotion
plt.figure(figsize=(10,6))
cmap=df.corr()
sns.heatmap(cmap)
```

Define X and y

```python
target = df.play

df.drop(['play', 'day'], axis = 1, inplace = True)
df
```

```python
y = target
X = df
```

Split for training and validation

```python
rows = int(len(df) * .7)

X_train = X[: rows]
y_train = y[: rows]

X_val = X[rows :]
y_val = y[rows :]

X_train.shape, y_train.shape, X_val.shape, y_val.shape
```

Define model

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

model = GaussianNB().fit(X_train, y_train)
print(model.score(X_train, y_train))
```

Predict on validation set

```python
y_pred = model.predict(X_val)
print(accuracy_score(y_val, y_pred))
```
