* * * * **default and flexible argumants**

```python
def d_arg(a,b,c=100):
    x=a*b*c
    return x
print(d_arg(3,4))  # c variable value is to be default 100
print(d_arg(3,4,5))  # c variable value is to be given value 5
```

**flexible argument**

```python
def square_1(*args):
    for i in args:
        i*=i
        print(i)
square_1(3,4,5)    #  number of variables sent is changeable
```

```python
def f(**kwargs):    # key value arguments
    for k,v in kwargs.items():
        print(k,' ',v)
f(ali=35,veli=45,hasan=25,alperdede=75,total='hundred and eighty')
    
```
