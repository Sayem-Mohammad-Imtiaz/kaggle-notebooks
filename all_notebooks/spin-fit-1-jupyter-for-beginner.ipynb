### Shortcuts in Jupyter

* Press ESC to enter command mode and press ENTER to modify the cell. The commands below operate in command mode.

* A adds a new cell above the current cell.  
* B adds a new cell below the current cell.  
* Pressing D + D, D twice in succession deletes the current cell.  
* M converts to Markdown cell, Y converts to Code cell and R converts to Raw Cell.
* You can use the CTRL + B screen larger. The left file explorer disappears.
* SHIFT + M Merge two cells into one cell.  
* CTRL + SHIFT +- Separates into two cells based on the current cursor position.  
* SHIFT + J or SHIFT + DOWN Selects a new cell from the current cell to the lower position.
* SHIFT + K or SHIFT + UP Selects a new cell from the current cell to the top position.
* CTRL + / Comment out the selected code.
* CTRL + D Shortcut to delete one line.
* CTRL + [ or CTRL + ] Group indentation / Tab and Shift + Tab are also available,   
  but in text editor, Shift+Tab is not available, so CTRL + [ is useful.
