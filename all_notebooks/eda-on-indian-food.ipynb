## 04. Let's do some Analysis:

### 01. STATE

```python
print("STATES WITH NUMBER OF DISHES")
pd.crosstab(df["state"],["name"])
```

```python
top_5=pd.crosstab(df["state"],["name"]).sort_values(by=["name","state"],ascending=False)
top_5.head(5).plot(kind="bar",color="pink",figsize=(8,8))
plt.title("TOP 5 STATES WITH MAX NUMBER OF DISHES")
plt.show()
```

```python

pd.crosstab(df["state"],df["diet"]).plot(kind="bar",figsize=(8,8))
plt.title("STATE WISE MOST PREFFERD TYPE OF DIET")
plt.show()
```

```python
pd.crosstab(df["state"],df["course"]).plot(kind="bar",figsize=(8,8))
plt.title("STATE WISE MOST PREFFERD COURSE")
plt.show()
```

```python
pd.crosstab(df["state"],df["flavor_profile"]).plot(kind="barh",figsize=(8,8))
plt.title("STATE WISE MOST PREFFERD FLAVOR")
plt.show()
```
