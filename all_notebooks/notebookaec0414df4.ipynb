```python
plt.figure(figsize=(20,7))
sns.countplot(df['Country'])
```

```python
plt.figure(figsize=(20,7))
plt.style.use("fivethirtyeight")
sns.lineplot(x='Country',y='Goals',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,7))
plt.style.use("fivethirtyeight")
sns.lineplot(x='Country',y='Minutes played',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,7))
plt.style.use("fivethirtyeight")
sns.lineplot(x='Position',y='Goals',data=df)
plt.show()
```

```python
plt.figure(figsize=(20,7))
sns.barplot(x='Player',y='Minutes played',data=df)
plt.xticks(rotation=90)
plt.show()
```
