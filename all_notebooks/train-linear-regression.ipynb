```python
sales['price_th'] = sales['price'] / 1000
sales[['price', 'price_th']].head()
```

```python
sns.regplot(x = sales['sqft_living'], y = sales['price_th'])
```

```python
model = smf.ols('price_th ~ sqft_living', data=sales)
res = model.fit()
res.summary()
```

```python
sales.columns
```

```python

x = sales.loc[:, ['bedrooms','bathrooms','floors', 'sqft_living', 'grade', 'yr_built', 'view', 'sqft_lot', 'waterfront', 
                 'condition', 'sqft_above']]
y = sales.loc[:, 'price_th']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = LinearRegression().fit(x_train, y_train)
model.score(x_train, y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

The coefficient of determination (0.65) is higher than in model with one feature (above with sqft_living = 0.49)

```python
y_pred = model.predict(x_test)
```

```python
y_test = np.exp(y_test)
```

```python
# y_test = np.exp(y_test)
y_pred = np.exp(y_pred)
```

```python
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print('mse: %.3f, mae: %.3f' % (mse, mae)
```
