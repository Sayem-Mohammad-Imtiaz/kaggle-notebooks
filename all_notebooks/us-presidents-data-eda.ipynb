```python
import pandas as pd
from pandas_profiling import ProfileReport
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("../input/historical-us-president-physical-data-more/Historical Presidents Physical Data (More).csv")
```

```python
profile = ProfileReport(df, title='Pandas Profiling Report')
```

```python
profile
```

```python
plt.figure(figsize=[20,5])
base_color=sns.color_palette()[0]
order=df['birth_state'].value_counts().index
sns.countplot(data=df,x='birth_state',order=order,color=base_color)
plt.title('Birth State of US Presidents',fontsize=19,fontname='Times new roman')
plt.xticks(rotation=90);
```

- 8 Presidents came from virginia

```python
df[df['birth_state']=='Virginia']
```

```python
plt.figure(figsize=[10,4])
plt.hist(df['presidency_begin_age']);
plt.title('Presidency begin Age',fontname='Times new roman',fontsize=17);
plt.ylabel('Frequency');
plt.xlabel('Age');
```

- The most frequent age for being a president is around 55

```python
plt.figure(figsize=[20,5])
base_color=sns.color_palette()[0]
order=df['political_party'].value_counts().index
sns.countplot(data=df,x='political_party',order=order,color=base_color);
plt.title('Parties of US Presidents',fontsize=19,fontname='Times new roman');
```

- Most Presidents come from the Republican party followed by the Democrat Party

```python
plt.figure(figsize=[20,5])
base_color=sns.color_palette()[0]
order=df['body_mass_index_range'].value_counts().index
sns.countplot(data=df,x='body_mass_index_range',order=order,color=base_color);
plt.title('BMI of US Presidents',fontsize=19,fontname='Times new roman');
```

```python
plt.figure(figsize=[20,5])
my_colors=['AntiqueWhite','Beige','BlanchedAlmond','DarkGray','LightBlue']
my_explode = (0, 0.05, 0.05,0.05,0.05)
df['body_mass_index_range'].value_counts(normalize=True).plot(kind='pie',
                                                           fontsize=12,autopct='%1.1f%%', startangle=15, shadow = True
                                                           ,colors=my_colors,explode=my_explode)
plt.title('BMI classification of US presidents',fontname='Times new roman',fontsize=14)
plt.show()
```

- 37.8% of US presidents are overweight

```python
plt.figure(figsize=[10,4])
plt.hist(df['height_cm']);
plt.title('Distribution of US presidents Heights',fontname='Times new roman',fontsize=15);
plt.xlabel('Height(cm)');
plt.ylabel('Frequency');
```

```python
plt.figure(figsize=[10,4])
sns.boxplot(data=df,x='height_cm');
plt.title('Distribution of US presidents Heights',fontname='Times new roman',fontsize=15);
```

```python
plt.figure(figsize=[10,4])
sns.boxplot(data=df,x='height_cm',y='political_party');
```
