```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
StandardScaler = StandardScaler()
columns_to_scale=['age','creatinine_phosphokinase','ejection_fraction','platelets','serum_sodium','time']
dataset[columns_to_scale] = StandardScaler.fit_transform(dataset[columns_to_scale])
```

```python
dataset.head()
```

```python
y = dataset['DEATH_EVENT']
x = dataset.drop(['DEATH_EVENT'], axis = 1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import cross_val_score
knn_score=[]
for k in range(1,21):
    knn_classifier = KNeighborsClassifier(n_neighbors = k)
    score = cross_val_score(knn_classifier,x,y,cv=10)
    knn_score.append(score.mean())
```

```python
plt.plot([K for K in range(1,21)],knn_score, color ='red')
for i in range(1,21):
    plt.text(i,knn_score[i-1],(i,knn_score[i-1]))
plt.xticks([i for i in range(1,21)])
plt.xlabel('no. of neighbor')
plt.ylabel('scores')
plt.title('score')
```

```python
knn_classifier = KNeighborsClassifier(n_neighbors = 17)
score = cross_val_score(knn_classifier,x,y,cv=10)
```

```python
score.mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
randomforest_classifier = RandomForestClassifier(n_estimators= 17)
score = cross_val_score(randomforest_classifier,x,y,cv =10)
```
