```python
import pandas as pd
df=pd.read_csv('/kaggle/input/height-and-weight/htwt.csv')
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['Weight'],df['Height'])
plt.show()
```

```python
X=df.values
```

```python
# check for K value
# Elbow method

from sklearn.cluster import KMeans
sse = []
k = range(1,13)
for i in k:
    model = KMeans(n_clusters=i,random_state=0)
    model.fit(X)
    sse.append(model.inertia_)

plt.scatter(k,sse)
plt.plot(k,sse)
```

```python
from sklearn.cluster import KMeans
model=KMeans(n_clusters=2,random_state=0)
model.fit(X)
```

```python
y=model.predict(X)
y
```

```python
model.cluster_centers_
```
