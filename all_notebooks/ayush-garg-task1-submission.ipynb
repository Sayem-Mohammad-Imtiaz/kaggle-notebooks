```python
df['Murder']=df['child_victims']+df['male_vic']+df['female_vic']
df['kidnapping']=df['kidnapped_child_vic']+df['affected_male']+df['affected_female']
df['crime_women']=df['adult_victims']+df['child_involved']
total_crime=np.array(df.groupby(['city']).sum())
viz1=pd.DataFrame({'Murder':total_crime[:,8],
                   'kidnapping':total_crime[:,9],
                   'crime_woman':total_crime[:,10]},index=['Ghaziabad', 'Kanpur', 'Lucknow'])

viz1.plot.bar(rot=15, title="Different natures of crime citywise plotting",figsize=(8,10));
```

We can easily see that kidnapping is the major crime in all of this cities. 
Also, crime against women is considerably high in Ghaziabad than in Kanpur or Lucknow.
Otherwise all the three citie show similar trend.
