# SVM

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#import the dataset

data=pd.read_csv('../input/zoodataset/datasets_586_1114_zoo.csv')
```

```python
class_1=data[data['class_type']==1]
```

```python
#x and y
x=data.drop('class_type',axis=1)
y=data['class_type']
```

```python
#splitting the data

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#svm

from sklearn.svm import SVC

model=SVC()

model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
#accuracy score

from sklearn.metrics import accuracy_score

accuracy_score(y_pred,y_test)
```

# we got 100% accuracy

```python
#Implement K_fold validation

from sklearn.model_selection import cross_val_score

svc_model=SVC()
```

```python
model_cv=cross_val_score(svc_model,x,y,cv=4)
model_cv
```

```python
model_cv.mean()
```

```python
#Changing the C parameter

svc_c = SVC(C=10)
model_cv=cross_val_score(svc_c,x,y,cv=4)
```

```python
model_cv
```

```python
model_cv.mean()
```

```python
#Now give some kernal value to SVM.
#Default kernal will be linear.
#To find we have linear or non-linear data

svc_c= SVC (C=10,kernel='linear')

model_cv= cross_val_score(svc_c,x,y,cv=4)

model_cv.mean()
```
