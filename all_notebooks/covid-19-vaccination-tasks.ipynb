Task 2 - Which Country Vaccinated most people

```python
total_vaccinations = pd.DataFrame(df.groupby('country').sum())
total_vaccinations.sort_values('total_vaccinations', ascending = False).head()
```

```python
total_vacc_sorted = total_vaccinations[107:].sort_values(by="total_vaccinations", ascending = False)
total_vacc_sorted
```

```python
top_countries = sns.barplot(y = total_vacc_sorted.index, x= total_vacc_sorted['people_vaccinated'])
#top_countries.set_xticklabels(top_countries.get_xticklabels(), 
                          #rotation=90, 
                          #horizontalalignment='right')
top_countries
```

Task3 - Highest % of vaccinations

```python
people_vacc_df = df[['country', 'total_vaccinations_per_hundred']]
people_vacc_df.head()
```

```python
people_vacc_grouped = people_vacc_df.groupby("country").mean()
people_vacc_grouped.head()
```

```python
pvg = pd.DataFrame(people_vacc_grouped.sort_values(by="total_vaccinations_per_hundred", ascending=False))
pvg.head()
```

```python
sns.barplot(y = pvg.index[:25], x=people_vacc_grouped['total_vaccinations_per_hundred'][:25])
```
