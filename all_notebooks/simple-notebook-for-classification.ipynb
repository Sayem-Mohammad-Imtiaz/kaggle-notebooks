```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

filepath = '../input/mushroom-classification'
df = pd.read_csv(filepath + '/mushrooms.csv');
```

# Pre-processing

## Converting categorical data to numerical

```python
from sklearn.preprocessing import LabelEncoder

for col in df.columns: # Consider each column (except 'y' column, we deal with that separately)
    
    # Generate labels for each value in column
    le = LabelEncoder();
    le.fit(df[col].unique());
    
    # Print labels
    print('\n' + col + ':');
    for val in df[col].unique():
        print(val, int(le.transform([val])));
        
    # Transform data    
    df[col] = le.transform(df[col]);
    
print('\n Sample of transformed data-frame:');

df.head()
```

```python
# randomly shuffle the data
df = df.sample(frac=1, random_state = 20);

# Divide the samples for training and testing
X_train = df.drop(['class'], axis = 1)[0:6499]; 
y_train = df['class'][0:6499];
X_test = df.drop(['class'], axis = 1)[6500:8123];
y_test = df['class'][6500:8123];
```

# Machine learning

## Modeling - using neural networks

```python
from sklearn.neural_network import MLPClassifier

# train model
nn = MLPClassifier().fit(X_train, y_train);

# evaluate model
print("Accuracy on train-set:", "%.2f" %(nn.score(X_train, y_train)*100), '%');
print("Accuracy on test-set:", "%.2f" %(nn.score(X_test, y_test)*100), '%');
```

## Modeling - using SVMs

```python
from sklearn.svm import SVC

# train model
svm = SVC(gamma = 'auto').fit(X_train, y_train);

# evaluate model
print("Accuracy on train-set:", "%.2f" %(svm.score(X_train, y_train)*100), '%');
print("Accuracy on test-set:", "%.2f" %(svm.score(X_test, y_test)*100), '%')
```

## Modeling - using Decision trees

```python
from sklearn import tree

# train model
dt = tree.DecisionTreeClassifier().fit(X_train, y_train);

# evaluate model
print("Accuracy on train-set:", "%.2f" %(dt.score(X_train, y_train)*100), '%');
print("Accuracy on test-set:", "%.2f" %(dt.score(X_test, y_test)*100), '%');
```
