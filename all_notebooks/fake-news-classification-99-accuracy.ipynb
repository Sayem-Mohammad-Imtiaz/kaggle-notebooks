```python
import pandas as pd
from sklearn import model_selection,linear_model,naive_bayes, ensemble, metrics
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
fake = pd.read_csv("../input/fake-and-real-news-dataset/Fake.csv",usecols=['text'])
fake['label'] = 0
fake.head()
```

```python
truth = pd.read_csv("../input/fake-and-real-news-dataset/True.csv",usecols=['text'])
truth['label'] = 1
truth.head()
```

```python
df = pd.concat([truth, fake])
df.head()
```

```python
xtrain,xtest,ytrain,ytest = model_selection.train_test_split(list(df.text),df.label,test_size = 0.2, stratify = df.label, shuffle =True)
```

```python
tfidf = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3), use_idf=1,smooth_idf=1,sublinear_tf=1,
            stop_words = 'english')

tfidf_text = tfidf.fit_transform(list(xtrain)+list(xtest))
tfidf_xtrain = tfidf.transform(xtrain)
tfidf_xtest = tfidf.transform(xtest)
```

```python
lr = linear_model.LogisticRegression().fit(tfidf_xtrain,ytrain)
preds = lr.predict(tfidf_xtest)
print(model_selection.cross_val_score(lr,tfidf_xtest,ytest,cv = 5))
print(metrics.accuracy_score(preds,ytest))
```
