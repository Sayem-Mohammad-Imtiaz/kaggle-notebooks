```python
#Codes by ask9 https://www.kaggle.com/arbazkhan971/indians-vs-world/notebook

def random_colors(number_of_colors):
    color = ["#"+''.join([random.choice('0123456789ABCDEF') for j in range(6)])
                 for i in range(number_of_colors)]
    return color
```

```python
from pywaffle import Waffle
import random

death = df["death_rate_per_100,000_fully_vaccinated"].value_counts()

fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=20,
    values=death,
    colors = random_colors(len(death)),
    title={'label': 'Death per 100,000 fully Vaccinated ', 'loc': 'left'},
    labels=["{}({})".format(a, b) for a, b in zip(death.index, death) ],
    legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.4), 'ncol': len(df.groupby("month").get_group('May')), 'framealpha': 0},
    font_size=30, 
    icons = 'child',
    figsize=(12, 5),  
    icon_legend=True
)
```

```python
non_vaccinated = df["death_rate_per_100,000_not_fully_vaccinated"].value_counts()

fig = plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=20,
    values=non_vaccinated,
    colors = random_colors(len(non_vaccinated)),
    title={'label': 'Not fully vaccinated: deaths per 100,000', 'loc': 'left'},
    labels=["{}({})".format(a, b) for a, b in zip(non_vaccinated.index, non_vaccinated) ],
    legend={'loc': 'lower left', 'bbox_to_anchor': (0, -0.4), 'ncol': len(df.groupby("month").get_group('May')), 'framealpha': 0},
    font_size=30, 
    icons = 'child',
    figsize=(12, 5),  
    icon_legend=True
)
```

```python
fig = px.bar(df, 
             x='month', y='death_rate_per_100,000_not_fully_vaccinated',color_discrete_sequence=['blue'],
             title='Death rate per 100,000 not fully vaccinated', text='Percent_ people_completed_vaccine_by_first_of_the_month')
fig.show()
```

<h1><span class="label label-default" style="background-color:black;border-radius:100px 100px; font-weight: bold; font-family:Garamond; font-size:20px; color:#03e8fc; padding:10px">Breakthrough Infections</span></h1><br>

"Fully vaccinated people who test positive for COVID-19 more than 2 weeks after their completed vaccine dose series are called breakthrough infections. Breakthrough infections shown in the data only include those that meet the Centers for Disease Control (CDC) vaccine breakthrough infection definition(link is external). They identified vaccine breakthrough infections by comparing immunization records in WIR to confirmed and probable case records in WEDSS. This allows them to match person records between COVID-19 cases and vaccination status. The non-fully-vaccinated population includes individuals who:" 

Have no COVID-19 vaccine doses reported in WIR.

Have an incomplete COVID-19 vaccine series reported in WIR.

Are within the two-week window after completing their COVID-19 vaccine series required to build full immunity

Data source: The Wisconsin Immunization Registry (WIR) and Wisconsin Electronic Disease Surveillance System (WEDSS).
