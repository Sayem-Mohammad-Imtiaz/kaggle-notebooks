```python
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import math
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVR
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
data=pd.read_csv("../input/flight-take-off-data-jfk-airport/M1_final.csv")
```

```python
data.head(10)
```

```python
le=preprocessing.LabelEncoder()
le_data=data
le_data['Condition']=le.fit_transform(le_data['Condition'])
le_data['Condition'].unique()
```

```python
le_data.columns=['c1','c2','c3','c4','c5','c6','c7','c8','c9','c10','c11','c12','c13','c14','c15','c16','c17','c18','c19','c20','c21','c22','c23']
le_data.head(5)
```

```python
le_data.drop('c1',inplace=True, axis=1)

```

```python
le_data.head()
```

```python
le_data.drop('c5',inplace=True, axis=1)
le_data.drop('c2',inplace=True, axis=1)
le_data.drop('c3',inplace=True, axis=1)
le_data.drop('c4',inplace=True, axis=1)
le_data.head(5)
```

```python
le_data.head(10)
```

```python
le_data['c6']=le.fit_transform(le_data['c6'])
le_data.head(5)
```

```python
le_data['c16']=str(le_data['c16'])
le_data['c16']=le.fit_transform(le_data['c16'])
le_data.head(5)
```

```python
X=le_data.iloc[:,:17]
Y=le_data.loc[:,'c23'] 
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, train_size= 0.9, random_state=1)  
```

```python
scale= StandardScaler()  
X_train= scale.fit_transform(X_train) 
X_test= scale.transform(X_test)  
```

```python
lreg=LinearRegression()
lreg.fit(X_train,Y_train)
pred=lreg.predict(X_test)
mse = sklearn.metrics.mean_squared_error(Y_test, pred)
rmse_le_lreg=math.sqrt(mse)
print(rmse_le_lreg)
```

```python
ridreg=Ridge(alpha=0.05, normalize=True)
ridreg.fit(X_train,Y_train)
pred=ridreg.predict(X_test)
mse = sklearn.metrics.mean_squared_error(Y_test, pred)
rmse_le_rreg=math.sqrt(mse)

print(rmse_le_rreg)
```

```python
lasreg=Lasso(alpha=0.05, normalize=True)
lasreg.fit(X_train,Y_train)

pred=lasreg.predict(X_test)

mse = sklearn.metrics.mean_squared_error(Y_test, pred)
rmse_le_lasreg=math.sqrt(mse)

print(rmse_le_lasreg)
```

```python
knn=KNeighborsClassifier(n_neighbors=100)
knn.fit(X_train,Y_train)
pred=knn.predict(X_test)
mse = sklearn.metrics.mean_squared_error(Y_test, pred)
rmse_le_knn=math.sqrt(mse)
print(rmse_le_knn)
```

```python
svr=SVR(kernel='rbf')
svr.fit(X_train,Y_train)
pred=svr.predict(X_test)
mse = sklearn.metrics.mean_squared_error(Y_test, pred)
rmse_le_svr=math.sqrt(mse)
print(rmse_le_svr)
```

```python
gnb=GaussianNB()
gnb.fit(X_train,Y_train)
pred=gnb.predict(X_test)
mse = sklearn.metrics.mean_squared_error(Y_test, pred)
rmse_le_gnb=math.sqrt(mse)
print(rmse_le_gnb)
```

```python
random=RandomForestClassifier()
random.fit(X_train,Y_train)
pred=random.predict(X_test)
mse = sklearn.metrics.mean_squared_error(Y_test, pred)
rmse_le_rfc=math.sqrt(mse)
print(rmse_le_rfc)
```

```python
rmse_le={rmse_le_lreg,rmse_le_rreg,rmse_le_lasreg,rmse_le_knn,rmse_le_svr,rmse_le_gnb,rmse_le_rfc}
```
