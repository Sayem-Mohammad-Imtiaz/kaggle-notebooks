# Task: Create a new data frame with 3 columns to have these values Month, Year, Number of Orders

### convert order data into data time (dt)

```python
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

### create a new data frame

```python
df2 = pd.DataFrame()
```

### add a Month column to the new dataframe

```python
df2['Month'] = df['Order_Date'].dt.month
```

### add a Day column to the new dataframe

```python
df2['Day']= df['Order_Date'].dt.day
```

### add a Year column to the new dataframe

```python
df2['Year']=df['Order_Date'].dt.year
```

### add a number of orders column to the new dataframe

```python
df2['Number of orders'] = df['Units_Sold']
```
