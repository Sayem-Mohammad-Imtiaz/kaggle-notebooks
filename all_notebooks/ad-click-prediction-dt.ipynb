# Now let's prepare the data for trainin

```python
from sklearn.model_selection import train_test_split
```

# One hot encoder for gender

```python
data['Gender'] =data['Gender'].replace({"Male":0,"Female":1})
```

```python
Y = data['Purchased']
X = data.drop(['Purchased'], axis=1)
```

```python
random_state = 2
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state= random_state)
```

# Preparing the model

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy', random_state=random_state, min_samples_leaf=30)
```

```python
clf.fit(x_train, y_train)
```

# Evaluate the model

```python
from sklearn.metrics import accuracy_score
```

```python
pred = clf.predict(x_test)
pred
```

```python
acc = accuracy_score(y_test, pred)
acc*100
```

```python
from sklearn.metrics import classification_report
```

```python
clf_rep = classification_report(y_test, pred)
clf_rep
```

```python
plt.figure(figsize=(10,10))
tree.plot_tree(clf) 
```

```python
import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("diabetes")
```

```python
dot_data = tree.export_graphviz(clf, out_file=None, 
                     feature_names=X.columns,  
                     class_names=["1","0"],  
                     filled=True, rounded=True,  
                     special_characters=True)  
graph = graphviz.Source(dot_data)  
plt.figure(figsize=(8,10))
graph.render("diabetes2")
```
