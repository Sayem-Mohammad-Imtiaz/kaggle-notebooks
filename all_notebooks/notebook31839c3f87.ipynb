Kuna andmete formaat oli graafiku tegemiseks ebasobiv, siis pidime muutma andmeid.

```python
df["uus"] = df["Gross National Income (GNI) per Capita"].str.replace(",", "")
df["uus"] = pd.to_numeric(df["uus"])
```

```python
df.groupby(["Expected Years of Education", "uus"]).count()
```

Sorteerisime riigid ära SKP (per Capita) järgi ära, et saada ülevaate, kas lahendasime andmete muutmise õigesti ära.

```python
df = df.sort_values("uus", ascending = False)
pd.DataFrame({"Riik" : df["Country"],
             "SKP" : df["uus"]})
```
