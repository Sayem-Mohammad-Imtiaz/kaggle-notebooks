#### As we can see, there are six columns and no non-null values. But the Date column is of the object type. Let us convert it to datetime type for our convenience. 

```python
stock_data= pd.read_csv('../input/apple-aapl-historical-stock-data/HistoricalQuotes.csv', parse_dates=['Date'],skipinitialspace=True)
stock_data.info()
```

```python
display(stock_data.head())
```

#### Since we want to perform numerical calculations on all columns showing stock price, we need to convert them to float first. 

```python
stock_data['Close/Last']= stock_data['Close/Last'].str.replace('$','').astype(float,errors='raise')
stock_data.head()
```

```python
stock_data['Open']= stock_data['Open'].str.replace('$','').astype(float,errors='raise')
stock_data['Low']= stock_data['Low'].str.replace('$','').astype(float,errors='raise')
stock_data['High']= stock_data['High'].str.replace('$','').astype(float,errors='raise')
```

##### Now let us check the data types of all columns

```python
stock_data.dtypes
```

#### Now let us start with the analysis of this data

### When did the price of Apple stock reach it highest value and lowest value?

```python
highest_stock_date= stock_data.loc[stock_data['High']==stock_data['High'].max()]
display(highest_stock_date)
lowest_stock_date= stock_data.loc[stock_data['High']==stock_data['High'].min()]
display(lowest_stock_date)
```

### Highest jump in stock price

```python
stock_jump= stock_data.drop(columns= ['Close/Last','Volume','Open','Low'])
stock_data['Difference']= stock_jump.diff()['High']
stock_data.head(15)
```

```python
stock_data[stock_data['Difference']== stock_data['Difference'].max()]
```

### Similarly we find the date when the price had the biggest fall in a day.

```python
stock_data['Highest_Fall']= stock_data['High']- stock_data['Low']
stock_data[stock_data['Highest_Fall']== stock_data['Highest_Fall'].max()]
```

```python
stock_data= stock_data.set_index(['Date'])
stock_data.head()

```

Here we changed the index and converted date to DateTime type. 

```python
stock_data.index
```

#### Now we can access records based on the date. For example, getting data for only January month of 2021 is easy. 

```python
stock_data['2020-01-31':'2020-01-01']
```

#### We can also find average price of the stock for any time period. Let us see what the average open stock price for the month of January 2020 is.

```python
stock_data.loc['2020-01'].Open.mean()
```

#### Next we see how the stocks performed in the quarterly period of 2020. 

```python
viz_data= stock_data.loc['2019']['High'].resample('Q').mean()
viz_data
```

```python
%matplotlib inline
viz_data.plot()
```

### So we can see that the stock price has gone up from Q1 to Q4 in the year 2019 with a sharp rise from Q3 to Q4.
