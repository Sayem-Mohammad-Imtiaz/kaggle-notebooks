```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('../input/WATCHLIST.csv', encoding='latin-1')
```

```python
data.head()
```

```python
plt.figure(figsize=(18, 8))
sns.distplot(data['IMDb Rating'],kde = True)
plt.show()
```

```python
plt.figure(figsize=(18, 8))
ax = sns.countplot(x="Title Type", data=data);
```

```python
plt.figure(figsize=(18, 8))
sns.distplot(data[['Your Rating']], hist=False, kde_kws={"label": "My Rating"})
sns.distplot(data[['IMDb Rating']], hist=False, kde_kws={"label": "IMDb Rating"})

plt.show();
```

# Time Series Analysis

```python
# https://machinelearningmastery.com/time-series-data-visualization-with-python/
```

```python
data_ = data.loc[data['Date Rated'].notnull()]
data_ = data_.set_index('Date Rated')
data_drama = data_[data_['Genres'] == 'Drama']
data_drama
```

```python
data_drama = data_drama['Your Rating']
plt.figure(figsize=(25, 8))
data_drama.plot(style='k.')
plt.show()
```

```python
# Create figure and plot space
fig, ax = plt.subplots(figsize=(25, 10))

# Add x-axis and y-axis
ax.bar(data_drama.index.values,
       data_drama.values,
       color='purple')

# Set title and labels for axes
ax.set(xlabel="Date",
       ylabel="Rating",
       title="Changes in rating over time")

# Rotate tick marks on x-axis
plt.setp(ax.get_xticklabels(), rotation=45)

plt.show()
```
