```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing

from plotly.offline import init_notebook_mode, plot
import plotly as py
import plotly.graph_objs as go
init_notebook_mode(connected=True)
```

```python
meta = pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
meta.head()
```

```python
race = pd.DataFrame(meta['gender'].groupby(meta['race/ethnicity']).value_counts())
race["count"] = race.iloc[:,-1:]
del race['gender']
race.reset_index(inplace=True)
del race['gender']
race
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x = race['race/ethnicity'].unique(),
    y = race.groupby(['race/ethnicity']).sum()['count'],
    name='male',
    marker_color='#516882'
))

fig.update_layout(
    title="Race/Ethnicity wise student distribution",
    xaxis_title="Race/Ethnicity group ",
    yaxis_title="Student Count",
)

fig.show()
```

```python
df = pd.DataFrame(meta['test preparation course'].groupby(meta['gender']).value_counts())
df["count"] = df.iloc[:,-1:]
del df['test preparation course']
df.reset_index(inplace=True)
df
```

```python
male = df[df['gender'] != 'female']
male.reset_index(inplace = True, drop = True)
male
```

```python
female = df[df['gender'] != 'male']
female.reset_index(inplace = True, drop = True)
female
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x = male['test preparation course'].unique(),
    y = male['count'],
    name='male',
    marker_color='#6b005e'
))

fig.add_trace(go.Bar(
    x= female['test preparation course'].unique(),
    y= female['count'],
    name='female',
    marker_color='#e65ed6'
))

fig.update_layout(
    title="Test preparation of students : gender-wise stats",
    xaxis_title="Preparation Status",
    yaxis_title="Student Count",
)


fig.update_layout(barmode='group', xaxis_tickangle=-90)
fig.show()
```
