# Metadata

```python
meta = pd.read_csv('metadata.csv')
print(meta.shape)
meta.head()
```

```python
meta.info()
```

```python
meta.describe()
```

## Examining the categorical varaibles

```python
for col in meta.columns:
    if(meta[col].dtype == 'O'):
        data = meta[col].fillna('NaN')
        unique, counts = np.unique(data, return_counts=True)
        if(len(unique) < 20):  
            size = [16,4]
            if(len(unique)==2): 
                size = [16,1]
            
            plt.figure(figsize=size)
            plt.title(col)
            plt.barh(unique, counts)
            plt.grid(axis='x')
            plt.show()
```

## Examining the Numerical Variables

```python
for col in meta.columns:
    data = meta[col]
    if(data.dtype == 'float64'):
        data = meta[col]
        plt.figure(figsize=[16,4])
        plt.plot(data)
        plt.title(col)
        plt.grid()
        plt.show()        
```

# Weather

```python
weather = pd.read_csv('weather.csv')
print(weather.shape)
weather.head()
```

```python
values, counts = np.unique(weather.site_id, return_counts=True)
plt.figure(figsize=[16,4])
plt.barh(values, counts)
plt.grid(axis='x')
plt.show()
```

```python
for col in weather.columns:
    if(col not in ['timestamp', 'site_id']):
        plt.figure(figsize=[16,5])
        plt.title(col)
        plt.plot(weather[col])
        plt.grid()
        plt.show()
```
