```python
list1=[]
n=int(input("enter number of list element:"))
print("input element:")
for i in range(n):
    t=int(input())
    list1.append(t)
    
temp=set(list1)
list2=list(temp)

for i in range(len(list2)):
    list2[i]=list2[i]**2
    
print(list2)
```
