# Density Heatmap for Crimes

```python

fig = px.density_mapbox(df, lat='latitude', lon='longitude', radius=5, zoom=7,
                        color_continuous_scale=px.colors.sequential.YlOrRd,
                        mapbox_style="stamen-terrain")
fig.update_layout(
        title = 'Mapbox Density Heatmap - Crimes',
)
fig.show()
```

# Animated Density Heatmap for Crimes

Data preparation to created the animated map using only the year

```python
df['time']
```

```python
df['date']=pd.to_datetime(df['time'], format='%Y-%m-%d').dt.year
```

#Sorting the dataframe by the variable date numeric values (years)

```python
df=df.sort_values('date')
```

```python
fig = px.density_mapbox(df, lat='latitude', lon='longitude', radius=5, zoom=7,
                        color_continuous_scale=px.colors.sequential.YlOrRd,
                        mapbox_style="stamen-terrain", animation_frame='date')
fig.update_layout(
        title = 'Mapbox Density Heatmap - Crimes',
)
fig.show()
```

# Animated Density Heatmap for Prejuizo (Loss)

```python
import plotly.express as px
fig = px.density_mapbox(df, lat='latitude', lon='longitude', z='valor_prejuizo',
                        radius=5, zoom=7,
                        color_continuous_scale=px.colors.sequential.YlOrRd,
                        mapbox_style="stamen-terrain", animation_frame='date')
fig.update_layout(
        title = 'Mapbox Density Heatmap - Prejuizo',
)
fig.show()
```
