```python
#importing libraries
import numpy as np
import pandas as pd
import sklearn
from matplotlib import pyplot as plt
%matplotlib inline
import math
```

```python
df=pd.read_csv("/kaggle/input/flight-take-off-data-jfk-airport/M1_final.csv")
df.head()
```

# Label Encoding

```python
# from sklearn.preprocessing import LabelEncoder
# label_enco=LabelEncoder()

# df['OP_UNIQUE_CARRIER']=label_enco.fit_transform(df['OP_UNIQUE_CARRIER']).astype(str)
# df['TAIL_NUM']=label_enco.fit_transform(df['TAIL_NUM']).astype(str)
# df['DEST']=label_enco.fit_transform(df['DEST']).astype(str)
# df['Wind']=label_enco.fit_transform(df['Wind']).astype(str)
# df['Condition']=label_enco.fit_transform(df['Condition']).astype(str)
# X=df.drop('TAXI_OUT',1)
# Y=df['TAXI_OUT']

from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()

df['OP_UNIQUE_CARRIER']=label.fit_transform(df['OP_UNIQUE_CARRIER'].astype(str)).astype(str)
df['TAIL_NUM']=label.fit_transform(df['TAIL_NUM'].astype(str)).astype(str)
df['DEST']=label.fit_transform(df['DEST'].astype(str)).astype(str)
df['Wind']=label.fit_transform(df['Wind'].astype(str)).astype(str)
df['Condition']=label.fit_transform(df['Condition'].astype(str)).astype(str)
X=df.drop('TAXI_OUT',1)
Y=df['TAXI_OUT']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.10)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
x=np.empty(7,dtype=float)
```

KNN Neighbor

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
KNR1=KNeighborsRegressor(n_neighbors=200)
KNR1.fit(X_train,Y_train)
y=KNR1.predict(X_test)
x[0]=mean_squared_error(Y_test,y)**0.5
print(x[0])
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression 
LR1=LinearRegression()
LR1.fit(X_train,Y_train)
y=LR1.predict(X_test)
x[1]=mean_squared_error(Y_test,y)**0.5
print(x[1])
```

Lasso Regression

```python
from sklearn.linear_model import Lasso
L1=Lasso(alpha=0.05,normalize=True)
L1.fit(X_train,Y_train)
y=L1.predict(X_test)
x[2]=mean_squared_error(Y_test,y)**0.5
print(x[2])
```

Ridge Regression

```python
from sklearn.linear_model import Ridge
R1=Ridge(alpha=0.05,normalize=True)
R1.fit(X_train,Y_train)
y=R1.predict(X_test)
x[3]=mean_squared_error(Y_test,y)**0.5
print(x[3])
```

Naive-Bayes 

```python
from sklearn.linear_model import BayesianRidge
BR1=BayesianRidge()
BR1.fit(X_train,Y_train)
y=BR1.predict(X_test)
x[4]=mean_squared_error(Y_test,y)**0.5
print(x[4])
```

SVR Model

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
S1=StandardScaler()
S2=StandardScaler()
copy1=X_train.copy()
copy2=X_test.copy()
copy1=S1.fit_transform(copy1)
copy2=S2.fit_transform(copy2)
s=SVR()
s.fit(copy1,Y_train)
y=s.predict(copy2)
x[5]=mean_squared_error(Y_test,y)**0.5
print(x[5])
```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
RFR=RandomForestRegressor()
RFR.fit(X_train,Y_train)
y=RFR.predict(X_test)
x[6]=mean_squared_error(Y_test,y)**0.5
print(x[6])
```
