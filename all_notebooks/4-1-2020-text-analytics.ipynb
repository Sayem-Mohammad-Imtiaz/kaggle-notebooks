```python
documents = ["cats say a moew the cats", "the dogs say woof", "dogs chase with cats"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vectorizer = CountVectorizer(stop_words="english")
```

```python
term_doc_matrix = count_vectorizer.fit_transform(documents)
```

```python
print(term_doc_matrix.toarray())
print(count_vectorizer.get_feature_names())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words="english")
```

```python
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)
```

```python
print(tfidf_matrix.toarray())
print(tfidf_vectorizer.get_feature_names())
```

```python
df = pd.read_csv("/kaggle/input/spam-or-not-spam-dataset/spam_or_not_spam.csv")
```

```python
df
```

```python
df["label"].value_counts()
```

```python
df.info()
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
tfidf_vectorizer = TfidfVectorizer(stop_words="english")
```

```python
X = tfidf_vectorizer.fit_transform(df["email"])
```

```python
X.toarray()
```

```python
y = df["label"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.toarray(), y, test_size=0.3)
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier()
```

```python
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
