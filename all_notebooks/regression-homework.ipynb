```python
x=df.pelvic_incidence.values.reshape(-1,1)

y=df.lumbar_lordosis_angle.values.reshape(-1,1)

plt.scatter(x,y)
plt.xlabel("pelvic_incidence")
plt.ylabel("lumbar_lordosis_angle")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

linear_reg=LinearRegression()

linear_reg.fit(x,y)

#b0=linear_reg.predict([[50]])
#print(b0)

#b1=linear_reg.coef_
#print(b1)
```

```python

y_head=linear_reg.predict(x)

plt.plot(x,y_head,color="red")
plt.scatter(x=x,y=y)
plt.show()
```

R2_score:Elimde olan veriyi fit ettigimizde dogruluk oranını bulmamızı saglar.
metrics:Degerlerndiren Kütüphane

```python
from sklearn.metrics import r2_score

print("R2 score:",r2_score(x,y_head))
```
