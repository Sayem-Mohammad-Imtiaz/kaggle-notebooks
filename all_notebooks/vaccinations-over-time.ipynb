```python
import pandas as pd
import numpy as np
from io import StringIO
import io
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../input/russia-covid-vaccination/Russia_Vaccination_Jun_week_4.csv", encoding="UTF-8", index_col='date', parse_dates=['date'])
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df = df.iloc[4:]
df.head()
```

```python
df["total_vaccinations"].asfreq('D', method='bfill').plot()
plt.title('Total vaccinations over time (Day frequency)')
plt.show()
```

```python
df.plot(subplots=True, figsize=(10,12))
plt.ylabel('amount of vaccinations')
plt.xlabel('date')
plt.title('Vaccinations over time')
plt.show()
```

```python
df['Ratio'] = df['people_fully_vaccinated'] / df['people_vaccinated'] * 100
```

```python
df['Ratio'].asfreq('D', method='bfill').plot()
plt.title('Percent of people who received all doses comparing to people first vaccinated')
plt.show()
```
