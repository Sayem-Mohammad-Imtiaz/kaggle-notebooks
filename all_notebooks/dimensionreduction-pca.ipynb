PCA en basit tanimiyla aralarindaki korelasyona gore oznitelik sayisini dusurerek verimi arttirmayi amaclar. Veriler arasindaki mesafeyi maksimize etmeye calisir. Verileri kÄ±yasladigi icin supervised learning ile kullanilabilir.

Kutuphanelerimizi ve verimizi dahil edelim.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../input/winewithclassification/Wine.csv')
```

```python
df.head()
```

Verimize bakacak olursak sarap iceriklerini goruyoruz.

```python
df.shape
```

14 oznitelikten olusan bir verimiz var. Bu oznitelik sayisini en az anlam kaybi ile azaltmayi amacliyoruz.

Verilerimizi ayiralim. Train test kullacanagiz.

```python
X = df.iloc[:, 0:13]
y = df.iloc[:, 13]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 0)
```

Train ve test olarak veri bolutmemizi yaptik.

```python
X_train.head()
```

Verileri bir skalaya sikistiralim.

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

Sira geldi oznitelik sayimizi dusurmeye. Modelimizi olusturalim.

```python
from sklearn.decomposition import PCA
```

Bir nesne olusturup fitleyelim.

```python
pca = PCA(n_components = 2) # kac oznitelik alinacagi
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
X_train_pca[5]
```

Goruldugu uzere artik 2 ozniteligimiz var. Train ve train2 degiskenlerini ayri ayri tahmin ettirip skorlara bakalim.

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
y_pred = classifier.predict(X_test)# 13 oznitelik ile tahmin.
```

```python
classifier_pca = LogisticRegression(random_state=0)
classifier_pca.fit(X_train_pca,y_train)
y_pred_pca = classifier_pca.predict(X_test_pca)# 2 oznitelik ile tahmin
```

Karsilastirmada kesinlik skorunu kullanmak iyi olacaktir.

```python
from sklearn.metrics import accuracy_score
```

```python
PCAOncesi = accuracy_score(y_test,y_pred)
PCASonrasi = accuracy_score(y_test,y_pred_pca)
```

```python
print("Onceki skor :  " + str(PCAOncesi))
print("Sonraki skor :  " + str(PCASonrasi))
```
