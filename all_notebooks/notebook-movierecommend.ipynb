```python
list1 = pd.read_csv('/kaggle/input/movies-dataset-1970-to-2020/movie_list_1.csv')
list2 = pd.read_csv('/kaggle/input/movies-dataset-1970-to-2020/movie_list_2.csv')
print(list1.shape)
print(list1.isnull().sum())
print('\n')
print(list2.shape)
print(list2.isnull().sum())
```

```python
list1.dropna(inplace=True)
print(list1.shape)
list1.isnull().sum()
```

```python
list1['full_string'] = list1['director']+' '+list1['cast_new']+' '+list1['genres_new']
list2['full_string'] = list2['director']+' '+list2['cast_new']+' '+list2['genres_new']
```

```python
list1.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
list2.head()
```

```python
cv = CountVectorizer()
matrix = cv.fit_transform(list1['full_string'])
cosine_mat1 = cosine_similarity(matrix)
cosine_mat1
```

```python
cv1 = CountVectorizer()
matrix1 = cv1.fit_transform(list2['full_string'])
cosine_mat2 = cosine_similarity(matrix1)
cosine_mat2
```

```python
print(cosine_mat1.shape)
print(cosine_mat2.shape)
```

```python
cosine_df1 = pd.DataFrame(cosine_mat1)
cosine_df2 = pd.DataFrame(cosine_mat2)
```
