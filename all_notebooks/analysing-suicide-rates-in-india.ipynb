```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Reading data

```python
df = pd.read_csv("/kaggle/input/suicides-in-india/Suicides in India 2001-2012.csv")
df.head(5)
```

**There are 237,519 rows and 7 columns.**

```python
df["Type_code"].value_counts()
```

# Dropping total (all india/states/uts) rows, there are 312 rows of each

```python
df["State"].value_counts()
```

```python
index_all_india = df[df["State"] == 'Total (All India)' ].index
df.drop(index_all_india, inplace = True)
```

```python
index_all_states = df[df["State"] == 'Total (States)' ].index
df.drop(index_all_states, inplace = True)
```

```python
index_all_ut = df[df["State"] == 'Total (Uts)' ].index
df.drop(index_all_ut, inplace = True)
```

# Which state has the highest number of suicide rates?

```python
state_df = df.groupby(['State'])
result = state_df['Total'].sum().sort_values()
result
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(1,1, figsize = (10,10))
ax.barh(result.index,result.values)
ax.set_xlabel("No. of suicids")
ax.ticklabel_format(style='plain') #to deal with scientific notation
```

**Maharashtra has the highest number of suicide rates, followed by West Bengal and Tamil Nadu**

# Analysing suicide rates in Maharashtra

```python
mh_df = state_df.get_group('Maharashtra')
mh_df
```

# Which year has the highest number of suicide rates?

```python
result = mh_df.groupby(['Year'])['Total'].sum()
result
```

```python
year = range(2001,2013)
plt.style.use('ggplot')
fig, ax = plt.subplots(1,1, figsize = (10,5))
ax.bar(result.index, result.values)
ax.set_ylabel("No. of suicides")
ax.set_xticks(year)
```

```python
plt.figure(figsize = (10,5))
plt.plot(result.index, result.values)
plt.xticks(year, rotation = 45)
plt.ylabel("No. of suicides")
plt.tight_layout()
plt.show()
```

**Year 2012 has the highest number, 80,560 suicides.
Year 2009 has the lowest number of suicides, 71,500**

# Cause of suicides

```python
filt = mh_df["Type_code"] == 'Causes'
mh_df.loc[filt].groupby('Type')['Total'].sum().sort_values(ascending = False)
```

**Most suicides were because of family problems.**

# Social status

```python
filt = mh_df["Type_code"] == 'Social_Status'
result = mh_df.loc[filt].groupby('Type')['Total'].sum().sort_values(ascending = False)
result
```

**Most suicides were among married people.**

```python
explode = [0.1,0,0,0,0]
plt.pie(result.values, explode = explode, shadow= True, startangle = 90,radius=1.2)
plt.tight_layout()
plt.legend(result.index, loc = 2, bbox_to_anchor=(1, 0., 0.5, 0.5))
plt.show()
```

```python
filt = mh_df["Type"] == 'Married'
mh_df.loc[filt].groupby('Gender')['Total'].sum()
```

**Married men tend to suicide more than women**

# Education status

```python
filt = mh_df["Type_code"] == 'Education_Status'
result = mh_df.loc[filt].groupby('Type')['Total'].sum().sort_values()
result
```

**This stat is very interesting:**  
  **People with no education have lower rates of suicide than people with half education. **

```python
plt.style.use("seaborn")
plt.barh(result.index,result.values)
plt.xlabel("No. of suicides")
plt.show()
```

# Professional status

```python
filt = mh_df["Type_code"] == 'Professional_Profile'
result = mh_df.loc[filt].groupby('Type')['Total'].sum().sort_values()
result
```

This stat is more disheartening, suicide rates among farmers is much higher than any other profession.

```python
plt.style.use("seaborn")
plt.barh(result.index,result.values)
plt.xlabel("No. of suicides")
plt.show()
```

# Methods used for suicide

```python
filt = mh_df["Type_code"] == 'Means_adopted'
result = mh_df.loc[filt].groupby('Type')['Total'].sum().sort_values()
result
```

```python
plt.style.use("seaborn")
plt.barh(result.index,result.values)
plt.xlabel("No. of suicides")
plt.show()
```
