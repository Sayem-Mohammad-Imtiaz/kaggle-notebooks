```python
# Breve análise dos dados. Dataset escolhido: anv.csv
df = pd.read_csv('../input/dataviz-facens-20182-aula-1-exerccio-2/anv.csv', delimiter=',')
df.head(15)
```

```python
# Vamos, agora, classificar as variáveis

qualitativaNominal = 'Qualitativa Nominal'
quantitativaDiscreta = 'Quantitativa Discreta'

colunas = ['Variáveis', 'Classificação']

variaveis = [
    ['aeronave_operador_categoria', qualitativaNominal],
    ['aeronave_tipo_veiculo', qualitativaNominal],
    ['aeronave_motor_tipo', qualitativaNominal],
    ['aeronave_assentos', quantitativaDiscreta],
    ['aeronave_ano_fabricacao', quantitativaDiscreta],
    ['aeronave_fase_operacao', qualitativaNominal],
    ['aeronave_tipo_operacao', qualitativaNominal],
    ['aeronave_nivel_dano', qualitativaNominal],
    ['total_fatalidades', quantitativaDiscreta],
    ['aeronave_dia_extracao', quantitativaDiscreta]
]

respostaAnv = pd.DataFrame(variaveis, columns=colunas)
respostaAnv
```

### Questão 1 - Item B - Tabela de frequência

Construa uma tabela de frequência para cada uma das **variáveis qualitativas** que você escolheu (caso não tenha escolhido nenhuma, deixe esta questão em branco). Uma dica: a função *value_counts()* do Pandas pode ser muito útil. =)


```python
# Realizar frequência de acordo com cada variável qualitativa
# Vimos que,diante da análise anterior, temos apenas Qualitativas Nominais

variaveisQualitativas = respostaAnv.loc[respostaAnv['Classificação'] == qualitativaNominal]['Variáveis']
variaveisQualitativas

for v in variaveisQualitativas:
    df_v = df[v].value_counts()
    print('Variável \t\t Frequência')
    print(df_v)
    print('TOTAL \t\t', df_v.sum())
    
    print('\n\n')
```

### Questão 1 - Item C - Representação Gráfica
Para cada uma das variáveis, produza um ou mais gráficos, usando matplotlib, que descreva seu comportamento / caracteristica. Lembre-se que estes gráficos precisam ser compatíveis com a classificação da variável. 

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Quantidade x Nivel de dano

dfAeronaveNivelDanoSoma = df['aeronave_nivel_dano'].value_counts()
quantidadesNivelDano = dfAeronaveNivelDanoSoma.values
colunasNivelDano = dfAeronaveNivelDanoSoma.index

plt.figure(figsize=(10,5))
plt.title("Quantidade de Níveis de Dano pelo Tipo")
sns.barplot(x=colunasNivelDano, y=quantidadesNivelDano, palette = 'bright')
```

```python
# Total de fatalidades por tipo de operação

dfAgrupado =  df[['aeronave_tipo_operacao','total_fatalidades']]

# pegar apenas as fatalidades registradas, ou seja, total maior que zero
dfAgrupado = dfAgrupado.loc[dfAgrupado['total_fatalidades'] > 0]

# agrupar pelo tipo da operação e somar
dfAgrupado = dfAgrupado.groupby(['aeronave_tipo_operacao']).count()

tipoOperacao = dfAgrupado.index
totalFatalidades = dfAgrupado['total_fatalidades'].values

plt.figure(figsize=(10,5))
plt.title("Total de Fatalidades por Tipo de Operação")
sns.barplot(x=totalFatalidades, y=tipoOperacao, palette = 'bright')
```

```python
# Total de fatalidades pelo tipo de motor

dfAgrupado =  df[['aeronave_motor_tipo','total_fatalidades']]

# pegar apenas as fatalidades registradas, ou seja, total maior que zero
dfAgrupado = dfAgrupado.loc[dfAgrupado['total_fatalidades'] > 0]

# agrupar pelo tipo do motor e somar
dfAgrupado = dfAgrupado.groupby(['aeronave_motor_tipo']).count()

tipoMotor = dfAgrupado.index
totalFatalidades = dfAgrupado['total_fatalidades'].values

plt.figure(figsize=(10,5))
plt.title("Total de Fatalidades pelo Tipo do Motor")
sns.barplot(x=totalFatalidades, y=tipoMotor, palette = 'bright')
```
