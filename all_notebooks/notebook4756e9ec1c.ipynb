```python

path="../input/ma06noise/"
path2="../input/mitbih-annotation-csv-file/"

labeled_files=glob.glob(path2+ "*.csv")   
labeled_files=sorted(labeled_files)

all_files=glob.glob(path+ "*.csv")
all_files=sorted(all_files)
all_files=([s for s in all_files if s != '../input/mitbih-database/102.csv'])
all_files=([s for s in all_files if s != '../input/mitbih-database/104.csv'])
print(all_files)
print(labeled_files)
```

```python
m=0
dfx=None


for i in range(len(all_files)):
    
    
    with open(all_files[0]) as in_file:
        reader = csv.reader(in_file,delimiter='\t')
        result = [[item for item in row if item != ''] for row in reader]
        df=pd.DataFrame(result)
    
    print(all_files[i],i)
    print(labeled_files[i],i)
    df2=pd.read_csv(labeled_files[i])
    ML2=df[1]
    

    print(all_files[i],i)
    print(labeled_files[i],i)
    #df2=pd.read_csv(labeled_files[i])
    #ML2=df[1]
    #ML2[ 322422]=-15.330
    
    ML2=pd.to_numeric(ML2,errors='coerce', downcast='float',)
    #ML2 =ML2.replace(' ','').replace(',','.').replace("âˆ’", "-")
    #print(ML2)
    #print(type(df[1][0]))

    ML2=(ML2-ML2.mean())/ML2.std() 
    NumAux=df2['NumAux']

    Sample=df2['Sample']
    Type=df2['Type']

    
    Num=np.zeros(shape=(len(NumAux)))
    
    temp = 'N'

    for i in range(0,len(NumAux)):
        if NumAux[i] !=  ' ' :
            temp = NumAux[i]
            #print(temp + "paici kicu ekta")
        else :
            NumAux[i] = temp
            #print(temp + "kicu nai")
            
    for n in range(0,len(NumAux)):


        if NumAux[n]=="N":
            Num[n]=0
        elif NumAux[n]=="AB":
            Num[n]=1;
        elif NumAux[n]=="AFIB":
            Num[n]=2;
        elif NumAux[n]=="AFL":
            Num[n]=3;
        elif NumAux[n]=="B":
            Num[n]=4;
        elif NumAux[n]=="MISSB":
            Num[n]=5;
        elif NumAux[n]=="T":
             Num[n]=6;
        elif NumAux[n]=="VFL":
            Num[n]=7;
        elif NumAux[n]=="VT":
            Num[n]=8;
        elif NumAux[n]=="SVTA":
            Num[n]=9;
        elif NumAux[n]=="IVR":
            Num[n]=10;
        elif NumAux[n]=="P":
            Num[n]=11;
        elif NumAux[n]=="SBR":
            Num[n]=12; 
        elif NumAux[n]=="NOD":
            Num[n]=13; 
        
    print(Num)       
        
        
    beat = np.zeros(shape=(round(len(Sample)/3),4050)) 
    
    m=0
    n=0
    while n < (len(Sample)-5):
    
        beat[m,1:len(ML2[Sample[n]:Sample[n+5]])+1]=(ML2[Sample[n]:Sample[n+5]])
        beat[m,0]=mode(Num[n:n+3])
        m=m+1
        n=n+3

    
    if(dfx is None):
        dfx = beat
    else:
        dfx = np.concatenate((dfx,beat))
    
```
