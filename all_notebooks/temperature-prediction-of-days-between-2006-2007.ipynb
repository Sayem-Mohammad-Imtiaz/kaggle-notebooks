# Import Library and Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler
from warnings import filterwarnings
from sklearn.metrics import confusion_matrix, accuracy_score

filterwarnings("ignore")
data =pd.read_csv("../input/szeged-weather/weatherHistory.csv")
```

# Deleting Some Features

```python
delete =["Summary","Precip Type","Apparent Temperature (C)","Loud Cover","Apparent Temperature (C)","Daily Summary"]
```

# Taking One Year of Data

```python
data=data.drop(delete,axis=1)
data =data.iloc[:8786]
```

# Creating Month and Day Properties

```python
data['Month']= data['Formatted Date'].str.split(" ").str[0].str.split("-").str[1]
data['Formatted Date']= data['Formatted Date'].str.split(" ").str[0].str.split("-").str[2]

print(data["Formatted Date"])
```

# Creating a New Data Set

```python
temperature=[]
humidity=[]
wind_speed=[]
wind_bearing=[]
visibility=[]
pressure = []

temperature =data.groupby(['Month','Formatted Date'])["Temperature (C)"].mean()
humidity =data.groupby(['Month','Formatted Date'])["Humidity"].mean()
wind_speed=data.groupby(['Month','Formatted Date'])["Wind Speed (km/h)"].mean()
wind_bearing=data.groupby(['Month','Formatted Date'])["Wind Bearing (degrees)"].mean()
visibility=data.groupby(['Month','Formatted Date'])["Visibility (km)"].mean()
pressure=data.groupby(['Month','Formatted Date'])["Pressure (millibars)"].mean()
```

```python
data_two = {"Temperature (C)":temperature
           ,"Humidity " :humidity,"Wind Speed (km/h)":wind_speed
,"Wind Bearing (degrees)":wind_bearing,"Visibility (km)":visibility,"Pressure (millibars)":pressure}


datam=pd.DataFrame(data_two)
print(datam)
```

# Exploratory Data Analysis

```python
print(datam.head())
print(datam.shape)
print(datam.isnull().sum())
print(datam.info())
print(datam.describe().T)
```

```python
plt.plot(datam["Temperature (C)"],datam["Wind Speed (km/h)"], 'o',color ='black')
plt.xlabel("Temperature (C)")
plt.ylabel("Wind Speed (km/h)")
plt.show()
```

# Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
lbe=LabelEncoder()
data["Month"]=lbe.fit_transform(data["Month"])
```

# Split Data

```python
y = datam["Temperature (C)"]
x=datam.drop(["Temperature (C)"],axis=1)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=1)

print("x train: ",x_train.shape)
print("x test: ",x_test.shape)
print("y train: ",y_train.shape)
print("y test: ",y_test.shape)
```

# Standardize

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

# Build Model 

```python
from sklearn import metrics
model = LinearRegression().fit(x_train,y_train)
print(model.score(x_train,y_train))
y_pred =model.predict(x_test)


print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
model = LinearRegression().fit(x_train,y_train)
print(model.score(x_train,y_train))
y_pred =model.predict(x_test)


print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
pred=pd.DataFrame(y_pred)

plt.figure(figsize = (18, 6))
index=y_test.reset_index()["Temperature (C)"]
ax=index.plot(label="original values")
ax=pred[0].plot(label = "predicted values")
plt.legend(loc='upper right')
plt.title("Test vs Pred")
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.show()
```
