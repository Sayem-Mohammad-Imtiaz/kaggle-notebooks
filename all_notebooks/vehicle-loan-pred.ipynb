Let us first change the datatypes whherever necessary

```python
comb['MobileNo_Avl_Flag'] = comb['MobileNo_Avl_Flag'].astype('object')
comb['Aadhar_flag'] = comb['Aadhar_flag'].astype('object')
comb['PAN_flag'] = comb['PAN_flag'].astype('object')
comb['VoterID_flag'] = comb['VoterID_flag'].astype('object')
comb['Driving_flag'] = comb['Driving_flag'].astype('object')
comb['Passport_flag'] = comb['Passport_flag'].astype('object')
comb['VoterID_flag'] = comb['VoterID_flag'].astype('object')

comb['Date.of.Birth'] = comb['Date.of.Birth'].astype('datetime64')
comb['DisbursalDate'] = comb['DisbursalDate'].astype('datetime64')
```

```python
comb[['AVERAGE_ACCT_AGE_years', 'AVERAGE_ACCT_AGE_mons']] = comb['AVERAGE.ACCT.AGE'].str.split(expand = True)
comb['AVERAGE_ACCT_AGE_years'] = [X.strip('yrs') for X in comb['AVERAGE_ACCT_AGE_years']]
comb['AVERAGE_ACCT_AGE_mons'] = [X.strip('mon') for X in comb['AVERAGE_ACCT_AGE_mons']]
comb[['AVERAGE_ACCT_AGE_years', 'AVERAGE_ACCT_AGE_mons']]   
comb[['AVERAGE_ACCT_AGE_years', 'AVERAGE_ACCT_AGE_mons']] = comb[['AVERAGE_ACCT_AGE_years', 'AVERAGE_ACCT_AGE_mons']].astype('int64')
comb['AVERAGE_ACCT_AGEyears'] = round(((comb['AVERAGE_ACCT_AGE_years']*12)+ (comb['AVERAGE_ACCT_AGE_mons']))/12.2)
```
