# Prepare analysis

```python
import numpy as np
import pandas as pd
```

```python
data_df = pd.read_csv("/kaggle/input/energy-balance-in-europe/energy_balance_eu.csv")
```

```python
nrg_bal_dict_df = pd.read_csv("/kaggle/input/energy-balance-in-europe/nrg_bal_dict.csv")
siec_dict_df = pd.read_csv("/kaggle/input/energy-balance-in-europe/siec_dict.csv")
```

# Merge data with dictionaries

```python
data_df = data_df.merge(nrg_bal_dict_df)
data_df = data_df.merge(siec_dict_df)
```
