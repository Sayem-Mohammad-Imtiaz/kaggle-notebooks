# **Churn-Model-ANN**

Churn Model prediction using Artificial Neural Networks

Scikit-learn library has been used for encoding, splitting of dataset and feature scaling.<br>
Keras library which is a combination of tensorflow and theano has been used in creating the Artificial Neural Network for the Churn Model.<br>

The above mentioned libraries are the basic requirements.<br>

Tunning has been applied to the model, using K-Cross Validation.<br>
The code was run on spyder, after installing theano, tensorflow and Keras libraries. <br>

# Importing Libraries

```python
import pandas as pd
```

# Loading the Data Set

```python
df2 = pd.read_csv('../input/churn-modelling-data/Churn_Modelling.csv')
```

# EDA

```python
df2.head(5)
```

```python
df2.head(-5)
```

```python
categorical_cols = ['Geography',	'Gender']
df2 = pd.get_dummies(df2, columns = categorical_cols, drop_first = True)
df2.head(5)
```

```python
print(df2.columns)
```

# Dropping Unwanted Columns

```python
df2 = df2.drop(['RowNumber', 'CustomerId', 'Surname'], axis = 1)
```

# Feature Selection

```python
featurelist = ['CreditScore', 'Age', 'Tenure',
       'Balance', 'NumOfProducts', 'HasCrCard', 'IsActiveMember',
       'EstimatedSalary', 'Geography_Germany', 'Geography_Spain',
       'Gender_Male']
```

```python
outputlist = ['Exited']
```

```python
X = df2[featurelist]
y = df2[outputlist]
```

# Model Selection And Sacling

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33)
Scaler = StandardScaler()
Scaler.fit(X_train)
train_x = Scaler.transform(X_train)
test_x = Scaler.transform(X_train)
```

# ANN Model Building

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers


# Initializing the ANN
ann = tf.keras.models.Sequential()

# Adding the input layer and the first hidden layer

ann.add(tf.keras.layers.Dense(units=6, input_dim=len(featurelist),activation='relu'))
 
# Adding the second hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))


# Adding the output layer
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))



# Compiling the ANN
ann.compile(optimizer ='adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

# Training the ANN on the Training set
ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

# Confusion Matrix And Accuracy Score

```python
import numpy as np
y_pred = ann.predict(X_test)
y_pred = pd.DataFrame(y_pred)
y_pred.columns = ['Pred_Val']

y_pred = y_pred['Pred_Val'].apply(lambda x : 0 if x<=0.5 else 1)

y_pred.head(5)
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)

print(cm)
```

```python
accu =accuracy_score(y_test, y_pred)
print(accu)
```
