```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (10, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
plt.figure(figsize=(10,4))
plt.xticks(rotation=20)
plt.title('Abortion rate according to age')
sns.barplot(x=df['age_of_woman'], y=df['abortion_rate'], palette="Set2");
```

```python
plt.figure(figsize=(16,6))

sns.scatterplot( data=df, x="abortion_rate", y="period", hue="age_of_woman");
```

#Observing 4 trends :

Rate of abortion in women in theirs early teens (11-14) and over 40 are quite stable.

Rate of women in their 30's is moderate changing

Rate is quite changing in period years for women 15-19 and 25-29 years old

Finnaly rate values that are most changing in this period time (2000-2018) are found in women rage 20-24

#How abortion rate evolved in period 2000-2018 ?

```python
year_rate=df.groupby("period")[["abortion_rate"]].sum()
year_rate
```

```python
plt.figure(figsize=(25,6))
plt.title('Abortion rate through the years')
plt.xticks(rotation=70)
sns.barplot(x='period', y='abortion_rate', data=df);
```

#The higher abortion rate was recorded in 2004

The wave of values has a tendency to slowly decrese : we can see that the last 3 years values are lower then the 3 initials ones.

```python
tab2000_df=df.loc[0:7]

plt.figure(figsize=(16,10))
plt.title('Representation of abortion in 2000 according to women age')

plt.pie(tab2000_df.abortion_rate, labels= tab2000_df.age_of_woman, autopct='%1.1f%%', labeldistance=1.1, textprops=dict(color="k"));
```

In 2000 women between 20 and 29 years old represent 51.8% of the population that get an abortion.

Almost 20% of abortion were of teens aged 15 to 19 years old.

```python
tab2018_df=df.loc[144:151]
```

```python
plt.figure(figsize=(16,10))
plt.title('Representation of abortion in 2018 according to women age')

plt.pie(tab2018_df.abortion_rate, labels= tab2018_df.age_of_woman, autopct='%1.1f%%', labeldistance=1.1, textprops=dict(color="k"));
```

In 2018 still women in their 20's represent almost half of population (49.3%) that gets an abortion.

Value for teens in age 15 to 19 discreased of 9% from 2000 to 2018.
We can observe instead that values for women 30-34 and 35-39 increased in 2018.

#Inferences and Conclusion

We can say that the age of 20-24 years is certainly the most critical with its high abortion values, followed closely by the 25-29 age group.
