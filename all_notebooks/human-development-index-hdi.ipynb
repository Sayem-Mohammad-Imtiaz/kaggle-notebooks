## Fill all mising values 

```python
hdi['HDI Rank']=hdi['HDI Rank'].fillna(round(hdi['HDI Rank'].mean(),1))
hdi['1990']=hdi['1990'].fillna(round(hdi['1990'].mean(),3))
hdi['1991']=hdi['1991'].fillna(round(hdi['1991'].mean(),3))
hdi['1992']=hdi['1992'].fillna(round(hdi['1992'].mean(),3))
hdi['1993']=hdi['1993'].fillna(round(hdi['1993'].mean(),3))
hdi['1994']=hdi['1994'].fillna(round(hdi['1994'].mean(),3))
hdi['1995']=hdi['1995'].fillna(round(hdi['1995'].mean(),3))
hdi['1996']=hdi['1996'].fillna(round(hdi['1996'].mean(),3))
hdi['1997']=hdi['1997'].fillna(round(hdi['1997'].mean(),3))
hdi['1998']=hdi['1998'].fillna(round(hdi['1998'].mean(),3))
hdi['1999']=hdi['1999'].fillna(round(hdi['1999'].mean(),3))
hdi['2000']=hdi['2000'].fillna(round(hdi['2000'].mean(),3))
hdi['2001']=hdi['2001'].fillna(round(hdi['2001'].mean(),3))
hdi['2002']=hdi['2002'].fillna(round(hdi['2002'].mean(),3))
hdi['2003']=hdi['2003'].fillna(round(hdi['2003'].mean(),3))
hdi['2004']=hdi['2004'].fillna(round(hdi['2004'].mean(),3))
hdi['2005']=hdi['2005'].fillna(round(hdi['2005'].mean(),3))
hdi['2006']=hdi['2006'].fillna(round(hdi['2006'].mean(),3))
hdi['2007']=hdi['2007'].fillna(round(hdi['2007'].mean(),3))
hdi['2008']=hdi['2008'].fillna(round(hdi['2008'].mean(),3))
hdi['2009']=hdi['2009'].fillna(round(hdi['2009'].mean(),3))
hdi['2010']=hdi['2010'].fillna(round(hdi['2010'].mean(),3))
hdi['2011']=hdi['2011'].fillna(round(hdi['2011'].mean(),3))
hdi['2012']=hdi['2012'].fillna(round(hdi['2012'].mean(),3))
hdi['2013']=hdi['2013'].fillna(round(hdi['2013'].mean(),3))
hdi['2014']=hdi['2014'].fillna(round(hdi['2014'].mean(),3))
hdi['2015']=hdi['2015'].fillna(round(hdi['2015'].mean(),3))
hdi['2016']=hdi['2016'].fillna(round(hdi['2016'].mean(),3))
```

#### Again check for  null values

```python
hdi.isnull().sum()
```

### Drop unwanted data

```python
hdi = hdi.drop(['Coverage','Country'],axis=1)
```

## select data(features) and target(label) 

```python
X = hdi.iloc[:,1:]
y =hdi.iloc[:,:1]
```

```python
print(f'The shape of X is {X.shape}\n\nThe shape of y is {y.shape}')
```

```python
X.columns
```

```python
y.columns
```

### Split train test data

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1,random_state=1)
```

## Model Fitting

```python
model= LinearRegression().fit(X_train,y_train)
```

```python
Pred = model.predict(X_test)
Pred
```

```python
mean_squared_error(y_test,Pred)
```

```python
mean_absolute_error(y_test,Pred)
```

```python
r2_score(y_test,Pred)
```

# Model Evaluation

```python
training = model.predict(X_train)
training_data = r2_score(y_train,training)
training_data
```

```python
testing  = model.predict(X_test)
testing_data = r2_score(y_test,testing)
testing_data
```
