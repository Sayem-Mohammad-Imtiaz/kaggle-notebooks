```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df_train = pd.read_csv('/kaggle/input/mobile-price-classification/train.csv')
df_test = pd.read_csv('/kaggle/input/mobile-price-classification/test.csv')
```

```python
df_train.head()
```

```python
df_train.isnull().sum()
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(verbosity=0)
```

```python
x = df_train.drop('price_range',axis=1)
y = df_train['price_range']
```

```python
xgb.fit(x,y)
```

```python
from xgboost import plot_importance
plot_importance(xgb)
```

```python
df_test.isnull().sum()
```

```python
id = df_test['id']
```

```python
df_test.drop('id',axis=1,inplace=True)
```

```python
prediction = xgb.predict(df_test)
```

```python
data = {'Phone_id':id,"Phone_price":prediction}
df = pd.DataFrame(data)
df.head()
```
