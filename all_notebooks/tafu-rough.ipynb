```python
import pandas as pd
covid = pd.read_csv('../input/novel-corona-virus-2019-dataset/covid_19_data.csv')
covid.head(2)
```

Prothome ekta function create korte hobe jeta ekta input nibe ar ekta output dibe.. 
input e jeta dibi oitar upor conditional statement apply korte hbe..

```python
def func(n):                  # ekhane 'n' hocche func er input
    if 0 <= n < 10 :
        return 'Green Zone' 
    elif 10 <= n < 20 :
        return 'Yellow Zone'
    else :
        return 'Red Zone'
```

nicher code er left side : ***covid['Zone']*** nam e kono column age theke DF e nai so eta notun kore create hobe..then right side e *covid* er pore dot diye jei data tui func e input hishebe dite chas seta likhbi..then arekta dot diye *map(func_name)* dite hobe... main kaj ta mainly map function e kore..

```python
covid['Zone'] = covid.Confirmed.map(func)
```

ar notun column na banaye ager column ei kahini korle dekh = er left side e *covid['confirmed']* likhsi..

```python
covid['Confirmed'] = covid.Confirmed.map(func)
```
