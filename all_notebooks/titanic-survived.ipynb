DATA CLEANING

```python
plt.figure(figsize=(14,8))
sns.boxplot(x="Pclass",y="Age",data=df,palette="winter")
```

we can see that there is realtionship betwwen passanger class and age.i.e higher class generally by older people.

```python
def imput_age(cols):
    Age=cols[0]
    Pclass=cols[1]
    
    if pd.isnull(Age):
        if Pclass==1:
            return 37
        elif Pclass==2:
            return 29
        else:
            return 24
    
    else:
        return Age
        
```

APPLY THE FUNCTION

```python
df['Age']=df[['Age','Pclass']].apply(imput_age,axis=1)
```

```python
sns.heatmap(df.isnull(),yticklabels=False,cmap='BuPu')
```

drop column cabin as it is not used.



```python
df.drop('Cabin',axis=1,inplace=True)
```

```python
df.head()
```

converting categorical features

```python
df.info()
```

```python
embarked=pd.get_dummies(df["Embarked"],drop_first=True)
sex=pd.get_dummies(df['Sex'],drop_first=True)
```

```python
df.head()
```

```python
df.drop(['PassengerId','Name','Ticket','Fare','Sex','Embarked'],axis=1,inplace=True)
```

```python
df=pd.concat([df,sex,embarked],axis=1)
```

```python
df.head()
```

**BUILDING A LOGESTIC REGRESSION MODEL**


```python
df_train=df.drop(['Survived'],axis=1)
```

```python
df_test=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_train,df_test,test_size=0.30)
```

**TRAINING AND PREDCTING**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel=LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
prediction=logmodel.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
accuracy=confusion_matrix(y_test,prediction)
```

```python
accuracy
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy=accuracy_score(y_test,prediction)
accuracy
```
