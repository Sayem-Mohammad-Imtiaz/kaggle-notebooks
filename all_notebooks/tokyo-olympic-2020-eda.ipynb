```python
# Reading Dataset
olympics = pd.read_csv('../../kaggle/input/2021-olympics-medals-in-tokyo/Tokyo Medals 2021.csv')
```

```python
# Taking a look
print('****'*10)
print(f'Dataframe has {olympics.shape[0]} Rows, {olympics.shape[1]} Columns')
print('****'*10)


# head
olympics.head(10).sort_values(by="Rank By Total",
                        ascending=True).style.set_properties(**{'background-color': '#F0F0F0',
                                    'color': '#222222',
                                    'border': '1.5px  solid black'}).bar(color='#FEC63D', 
                                                                         vmin=100_000, 
                                                                         subset=['Gold Medal']).bar(color='#C2C2C2', 
                                                                                                    vmin=100_000, 
                                                                                                    subset=['Silver Medal']).bar(color='#CE6A2E', 
                                                                                                                                    vmin=100_000, 
                                                                                                                                    subset=['Bronze Medal'])
```

# **<center><span style='font-family:Georgia'> <span style='background:#DAC7A7'> üìàExploratory Data Analysis_ </span>**
    
<center><img src="https://projects.fivethirtyeight.com/olympics-medal-count/images/torch-crop.png?v=31e3e6b2"></center>
    <a href="https://projects.fivethirtyeight.com/olympics-medal-count/images/torch-crop.png?v=31e3e6b2" rel=" noreferrer nofollow">Image Source</a>

```python
#plot
plt.figure(figsize=(27,15));
ax = sns.barplot(y ='Country',x='Total',data=olympics[:30],color=olympics_col[0], 
                      zorder=2, linewidth=0, orient='h', saturation=1, alpha=1)

ax.set_xlabel("Total",fontsize=20, weight='bold')
ax.set_ylabel("Country",fontsize=20, weight='bold')
ax.tick_params(labelsize=10, width=0.5, length=1.5)

plt.title("Top 30 Countries with Medals",size=20,weight='bold')  

for a in ax.patches:
    value = f'{a.get_width():.0f}'
    x = a.get_x() + a.get_width() + 0.60
    y = a.get_y() + a.get_height() / 1.8 
    ax.text(x, y, value, ha='left', va='center', fontsize=12, 
            bbox=dict(facecolor='none', edgecolor='black', boxstyle='round', linewidth=0.2))

  


#image
path = mpimg.imread('../input/font-worksans/medal-crop.png')
imagebox = OffsetImage(path , zoom=1.6)
xy = (0.5, 0.7)
ab = AnnotationBbox(imagebox, xy, frameon=False, pad=1, xybox=(100.5, 16))

ax.add_artist(ab)
ax.text(x = 92.5, y = 22.5, s = 'Best Performance', fontsize=22, weight = 'bold',color=olympics_col[0])
ax.text(x = 95.5, y = 23.5, s = 'From U.S.A', fontsize=22, weight = 'bold',color=olympics_col[0]);
```

```python
figure = plt.figure(figsize=(30,10))

A = plt.pie(olympics['Total'][:10], 
        labels=olympics['Country'][:10],
        startangle=90,
        labeldistance=1.15,
        pctdistance=0.8,
        autopct='%1.1f%%')



plt.title("Pie Chart of Top 10 Countries with Medals",size=20,weight='bold')
plt.show();
```

```python
A = olympics[:15].plot(kind="bar",
                   y =['Gold Medal','Silver Medal','Bronze Medal'],
                   x = 'Country',
                   figsize=(20,6),
                   color=['#FEC63D',
                          '#C2C2C2',
                          '#CE6A2E']).legend(loc='upper center',
                                             ncol=3);

plt.title("Top 10 Countries with Medals",size=20,weight='bold') ;
```

```python
font_path = "../input/font-worksans/WorkSans-VariableFont_wght.ttf"

cloud = WordCloud(width=1080, 
                  height=270,
                  font_path=font_path,
                  colormap = "Dark2",
                  background_color='white').generate(" ".join(olympics['Country'].astype(str)))
plt.figure(figsize=(22, 10))
plt.imshow(cloud)
plt.title("Country",size=20,weight='bold') 
plt.axis('off');
```
