```python
#Explore the dataset
#Coldest Christmas
coldest_index = df['Celcius Min Temp'].idxmin()
coldest = df.iloc[112,0]
coldest = coldest.year
coldest_temp = df['Celcius Min Temp'].min()
print("The coldest Christmas in Chicago was in:", coldest)
print("The minimal temperature in degrees Celcius measured was:", coldest_temp,"degrees Celcius.")
print(" ")

#Warmest Christmas
warmest_index = df['Celcius High Temp'].idxmax()
warmest = df.iloc[111,0]
warmest_temp = df['Celcius High Temp'].max()
warmest = warmest.year
print("The warmest Christmas in Chicago was in:", warmest)
print("The highest temperature in degrees Celcius measured was:", warmest_temp,"degrees Celcius.")
print(" ")

#Christmas with most snowfall
snow_index = df['Snow'].idxmax()
snow = df.iloc[79,0]
snow = snow.year
print("The snowiest Christmas in Chicago was in:",snow)
print(" ")

#Last White Christmas
bool_snow = last_white_christmas = df['Snow'] > 0.00
bool_snow_true = np.where(bool_snow)[0]
last_element = bool_snow_true[-1]
last_white_christmas = df.iloc[146,0].year
print("Last White Christmas was in:")
print(last_white_christmas)
```

```python
#Data Visualization
#Temperature in Chicago @Christmas
plt.figure(figsize=(15,10))
sns.lineplot(x=df['Year'],y=df['Celcius Min Temp'],data=df)
sns.lineplot(x=df['Year'],y=df['Celcius High Temp'],data=df)

plt.title("Chicago Christmas Weather: Lowest and Highest Temperature")
plt.xlabel("Year")
plt.ylabel("Temperature in Celcius")
```

```python
#Snowfall
sns.countplot(df['White Christmas'],data=df)

plt.title("Chicago Christmas Weather: White Christmas")
plt.show()
```

```python
#Binary Classification
from sklearn.linear_model import LogisticRegression

#drop White Christmas Not Defined values
white_christmas_data = df[df['White Christmas'] != 'Not Defined']

#check if values for 'not defined' have been dropped, only true and false should remain
white_christmas_data['White Christmas'].value_counts()
print(len(white_christmas_data))  #this is also the length of the dataset

#count other missing values 
missing_values = white_christmas_data.isnull().sum()
print(missing_values)
```

```python
#drop the missing values from precipitation
white_christmas_data = white_christmas_data[white_christmas_data.Precipitation.notna()]

#check if missing values have been dropped
missing_values_dropped = white_christmas_data.isnull().sum()
print(missing_values_dropped)
```

```python
white_christmas_data.head(5)
```

```python
#Example: Logistic Regression
features = ['Fahrenheit Min Temp', 'Fahrenheit High Temp','Precipitation']
x = white_christmas_data[features]
y = white_christmas_data['White Christmas']

snowfall = LogisticRegression()
snowfall.fit(x,y)
snow_predict = snowfall.predict(x)

#model score
model_score = snowfall.score(x,y)
print(model_score)
```
