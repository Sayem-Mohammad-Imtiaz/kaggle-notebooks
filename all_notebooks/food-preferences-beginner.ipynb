Let us import data

```python
df= pd.read_csv('../input/food-preferences/Food_Preference.csv')
```

```python
df
```

Convert Datetime

```python
df['Dates'] = pd.to_datetime(df['Timestamp']).dt.date
df['Time'] = pd.to_datetime(df['Timestamp']).dt.time
```

```python
df
```

```python
df.drop('Timestamp',axis=1)
```

check for null values

```python
sns.heatmap(df.isnull(),cmap = 'viridis')
```

```python
df['Gender'].replace(np.nan, 'Female',inplace  = True)
```

```python
df.isnull().sum()
```

```python
df["Food"].value_counts()
```

```python
df["Gender"].value_counts()
```

```python
sns.stripplot(df['Gender'], df['Age'],jitter=True,palette=['#F8D030',"#C03028"])
```

```python
plt.figure(figsize = (35,10))
sns.countplot(x='Dates',data = df,palette=['#98D8D8',"#EE99AC"])
plt.xticks(rotation=30)
```

Top 20 types of Countries

```python
import plotly.express as px #distribution according to countries
top_rated=df[0:20]
fig =px.sunburst(
    top_rated,
    path=['Nationality'])
fig.show()
```

Types of Food

```python
sns.set(rc={'figure.figsize':(20,10)})

sns.countplot('Food',data=df)
```

```python
sns.FacetGrid(df, hue="Dates", height=6) \
   .map(sns.kdeplot, "Age") \
   .add_legend()
```

Food preference of different countries

```python
sns.catplot(data=df, x="Age", y="Nationality", col="Food",palette='cubehelix')
```

```python
df[df["Nationality"]=="Indian"]["Food"].value_counts()[:20].plot(kind="barh",color="deepskyblue")
plt.title("Type of Food that Indians prefer")
```

```python
plt.figure(figsize=(12,6))
df[df["Dessert"]=="Yes"]["Nationality"].value_counts()[:10].plot(kind="bar",color="Sandybrown")
plt.title("Dessert craving counries",size=18)
```

```python
sns.violinplot("Juice", "Age", data=df,
               palette=["lightblue", "lightpink"])
plt.title("Types of Juices which different Age group prefer",size=18)
```

Types of Food which different Age group prefer

```python
g = sns.FacetGrid(df, col="Food") 
g.map(sns.distplot, "Age")
```

```python
import plotly.express as px
fig = px.sunburst(df, path=['Gender', 'Dessert', 'Food'], values='Age',)
fig.show()
```

```python
Color_type = ['#78C850',  
                    '#F08030',  
                    '#6890F0',  
                    '#A8B820',  
                    '#A8A878',  
                    '#A040A0',  
                    '#F8D030',  
                    '#E0C068',  
                    '#EE99AC',  
                    '#C03028',  
                    '#F85888',  
                    '#B8A038', 
                    '#705898',  
                    '#98D8D8',  
                    '#7038F8',  ]
```

```python
sns.swarmplot(x='Nationality', y='Age', data=df, 
              palette=Color_type)
plt.xticks(rotation=45)
plt.title("Age group by types of countries",size=18)
```

```python
df["Dessert"].value_counts()
```

```python
sns.FacetGrid(df, hue="Dessert", height=6) \
   .map(sns.kdeplot, "Age") \
   .add_legend()
plt.title("Age Group with Dessert preference",size=18)
```

```python
sns.lmplot(x='Dessert', y='Age', data=df,
           fit_reg=False, # No regression line
           hue='Food') 
plt.title("Dessert preference in type of Food",size=18)
```
