```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../input/101-simple-linear-regressioncsv/1.01. Simple linear regression.csv")
data.head()
```

**Dataset:**

In the battle between SAT scores and GPA, many students wonder which heavyweight matters most to your college application. The most obvious is that the SAT is a standardized test.
While your GPA compares you to the rest of your school, your SAT score compares you to the rest of the country. GPAs are not standard.

```python
x = data['SAT']
y = data['GPA']
```

```python
# x_matrix = x.values.reshape(84,1)
x_matrix = x.values.reshape(-1,1)

x_matrix.shape
```

```python
reg = LinearRegression()
reg.fit(x_matrix, y)
```

```python
data2 = pd.DataFrame(data=[1740,1760], columns=['SAT'])
data2
```

```python
reg.predict(data2)
```

```python
data2['predicted_GPA'] = reg.predict(data2)
data2
```

```python
plt.scatter(x,y)
yhat = reg.coef_ * x_matrix + reg.intercept_
fig = plt.plot(x, yhat, lw=4, c='red')
plt.xlabel('SAT')
plt.ylabel('GPA')
plt.show()
```
