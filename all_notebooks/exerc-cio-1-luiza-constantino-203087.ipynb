### Questão 1 - Item C - Representação Gráfica
Para cada uma das variáveis, produza um ou mais gráficos, usando matplotlib, que descreva seu comportamento / caracteristica. Lembre-se que estes gráficos precisam ser compatíveis com a classificação da variável. 

```python
import matplotlib.pyplot as plt
import numpy as np

def createBarChart(x, y, title, label_x, label_y, f = 'N', c='black'):
            
    plt.bar(x, y, label = label_x, color = '#A9A9A9')
    
    frame = plt.gca()
    #oculta borda superior e direita
    frame.spines['right'].set_visible(False)
    frame.spines['top'].set_visible(False)    
    if f != False:
        frame.get_yaxis().set_ticks([]) #oculta valores eixo y

    plt.xlabel(label_x)
    plt.ylabel(label_y)
    plt.title(title)
    
    #imprime rótulo de dados S/N
    if f != False:
        for i in np.arange(0,len(y)): #imprime rótulos nas barras
            plt.text(x[i],y[i]/2, '{:.2f}%'.format(y[i]) if f=='%' else '{:,}'.format(y[i]) , color=c , horizontalalignment='center') 
        
    fig= plt.gcf()
    fig.set_size_inches(15, 6) 
    
    if len(x) >= 10:
        plt.xticks(rotation=30)
        
    plt.show()
```

```python
generos = ['gen_feminino','gen_masculino','gen_nao_informado']
total = df[generos].sum()

x = total.index
y = total

createBarChart(x, y, 'Quantidade de Eleitores por Gênero', '', '')

print('\n')
fig1, ax1 = plt.subplots()
explode = (0.05, 0.05, 0) 
ax1.pie(y, labels=x, explode = explode,  autopct='%1.1f%%', shadow=True, startangle=90, colors=['#BC8F8F','#4682B4','#32CD32'])
ax1.axis('equal')  
fig= plt.gcf()
fig.set_size_inches(15, 6) 
plt.show()
```

```python
dfFaixaxEtaria = class_var[class_var['Variavel'].str.contains("f_")].Variavel

total = df[dfFaixaxEtaria].sum()

x = total.index
y = total

createBarChart(x, y, 'Quantidade de Eleitores por Faixa Etária', '', '')
#Calculando em %
y = total.div(total.sum())*100


plt.plot(x, y, 'g--')
frame = plt.gca()
frame.spines['right'].set_visible(False)
frame.spines['top'].set_visible(False)    
frame.get_yaxis().set_ticks([]) #oculta valores eixo y

for i in np.arange(0,len(y)): #imprime rótulos nas barras
    plt.text(x[i], y[i], '{:.2f}%'.format(y[i]), color='black', fontweight='bold') 
    
fig= plt.gcf()
fig.set_size_inches(15, 6) 
plt.show()
```

```python
total = df[['uf','total_eleitores']].groupby('uf').sum().reset_index().sort_values('total_eleitores', ascending=False)

x = total.uf
y = total.total_eleitores

createBarChart(x, y, 'Quantidade de Eleitores por Estado', '', '', False )
```

```python
import squarify # pip install squarify

total = df[['nome_municipio','total_eleitores']].groupby('nome_municipio').sum().sort_values('total_eleitores', ascending=False).head(10).reset_index()
#outros = df.total_eleitores.sum() - total.total_eleitores.sum()
#total = total.append(pd.DataFrame({'nome_municipio': 'Outros', 'total_eleitores': outros},index=[0]), ignore_index=False)

sizes = total.total_eleitores
label = total.nome_municipio

squarify.plot(sizes=sizes, label=label, alpha=0.6 )
    
fig= plt.gcf()
fig.set_size_inches(12, 8) 
plt.title('Top 10 Municípios com Maior número de Eleitores')    
plt.axis('off')
plt.show()

createBarChart(label, sizes, '', '', '')
```

```python
total = df[['nome_municipio','total_eleitores']].groupby('nome_municipio').sum().sort_values('total_eleitores', ascending=True).head(10).reset_index()
#outros = df.total_eleitores.sum() - total.total_eleitores.sum()
#total = total.append(pd.DataFrame({'nome_municipio': 'Outros', 'total_eleitores': outros},index=[0]), ignore_index=False)

sizes = total.total_eleitores
label = total.nome_municipio

squarify.plot(sizes=sizes, label=label, alpha=0.6 )
    
fig= plt.gcf()
fig.set_size_inches(12, 8) 
plt.title('Top 10 Municípios com Menor número de Eleitores')    
plt.axis('off')
plt.show()

createBarChart(label, sizes, '', '','')
```

```python
def stackedGen(total, label, title) :
    
    labels = total[label]
    men = total.gen_masculino
    women = total.gen_feminino
    na = total.gen_nao_informado
    width = 0.5       # the width of the bars: can also be len(x) sequence

    fig, ax = plt.subplots()

    ax.bar(labels, na, width, label='Não Informado', color = '#5bd75b')
    ax.bar(labels, men, width, bottom=na, label='Masculino', color = '#7da7ca')
    ax.bar(labels, women, width, bottom=men, label='Feminino', color = '#BC8F8F')

    frame = plt.gca()
    #oculta borda superior e direita
    frame.spines['right'].set_visible(False)
    frame.spines['top'].set_visible(False)        

    ax.legend()

    fig.set_size_inches(15, 6) 
    plt.title(title)
    plt.show()
```

```python
total = df[['uf', 'gen_feminino','gen_masculino','gen_nao_informado']].groupby('uf').sum().reset_index().sort_values('gen_masculino', ascending=False).head(15)
stackedGen(total, 'uf', 'Quantidade de Eleitores por Estado')
```

```python
dfFaixa = class_var[class_var['Variavel'].str.contains("f_")].Variavel
dfGen = class_var[class_var['Variavel'].str.contains("gen_")].Variavel

total = df[dfGen].groupby(dfFaixa).sum().reset_index()

stackedGen(total, 'Variavel', 'Quantidade de Eleitores por Faixa Etária')
```
