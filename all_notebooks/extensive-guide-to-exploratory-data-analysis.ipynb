## Data Description
   
- instant: record index

- dteday : date

- season : season (1:springer, 2:summer, 3:fall, 4:winter)

- yr : year (0: 2011, 1:2012)

- mnth : month ( 1 to 12)

- hr : hour (0 to 23)

- holiday : weather day is holiday or not

- weekday : day of the week

- workingday : if day is neither weekend nor holiday is 1, otherwise is 0.

- weathersit : 
    - 1: Clear, Few clouds, Partly cloudy, Partly cloudy

    - 2: Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist

    - 3: Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds

    - 4: Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog

- temp : Normalized temperature in Celsius. The values are divided to 41 (max)

- atemp: Normalized feeling temperature in Celsius. The values are divided to 50 (max)

- hum: Normalized humidity. The values are divided to 100 (max)

- windspeed: Normalized wind speed. The values are divided to 67 (max)

- cnt: count of total rental bikes including both casual and registered

```python
import pandas as pd # Library for Dataframe operations
import numpy as np
import matplotlib.pyplot as plt # Plotting library
import seaborn as sns   # Plotting library
```

<div class="alert alert-success" role="alert">
<h1><span class="label label-success">View at the dataset</span></h1>
    </div>

```python
train = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/bike_data/bike_train.csv" )

train.head(3).append(train.tail(3))
```

```python
train.info()
```

* ***Here we observe that there is no missing values in the dataset and we are good to go with preprocessing setps.***

```python
train['dteday'] = pd.to_datetime(train['dteday'], format = '%d-%m-%Y')
train_sort = train.sort_values(by = 'instant' , ascending= True,ignore_index= True)
train_sort.head()
```

* ***We convert the dates into date-time format and sort the entire data with respect to date***.

```python
train_sort.describe(percentiles = [.25,.5,.75,.95,.97])
```

* ***The summary of data provides us lot of information such as mean, variance,min ,max etc.***
* ***By view at the percentiles we get an idea of outliers in the data***

```python
# Renaming the column names
train_sort.rename(columns={'instant':'rec_id',
                        'dteday':'datetime',
                        'holiday':'is_holiday',
                        'workingday':'is_workingday',
                        'weathersit':'weather_condition',
                        'hum':'humidity',
                        'mnth':'month',
                        'cnt':'total_count',
                        'hr':'hour',
                        'yr':'year'},inplace=True)
```

```python
# Changing datatype of categorical columns 
train_sort['is_holiday'] = train_sort['is_holiday'].astype('category')
train_sort['is_workingday'] = train_sort['is_workingday'].astype('category')
train_sort['month'] = train_sort['month'].astype('category')
train_sort['hour'] = train_sort['hour'].astype('category')
train_sort['year'] = train_sort['year'].astype('category')
```

<div class="alert alert-success" role="alert">
<h1><span class="label label-success">   Count plots  </span></h1>
    </div>

```python
fig,ax = plt.subplots(2, 3, figsize=(17, 10), sharey=False)     


sns.countplot(x = train_sort['is_holiday'],ax=ax[0,0])
ax[0,0].set_title("Count plot of shares on holiday")
sns.countplot(x  = train_sort['season'], ax = ax[0,1])
ax[0,1].set_title("Count plot of shares different seasons")
sns.countplot(x  = train_sort['is_workingday'], ax = ax[0,2])
ax[0,2].set_title("Count plot of shares on working day")
sns.countplot(x  = train_sort['month'], ax = ax[1,0])
ax[1,0].set_title("Count plot of shares on months")
sns.countplot(x  = train_sort['year'], ax = ax[1,1])
ax[1,1].set_title("Count plot of shares on year")
sns.countplot(x  = train_sort['hour'], ax = ax[1,2])
ax[1,2].set_title("Count plot of shares on hours")


```

```python

column = ['temp','atemp','humidity','windspeed','total_count']

sns.pairplot(train_sort[column],height = 1.5)
```

<div class="alert alert-success" role="alert">
<h1><span class="label label-success">Regression plots </span></h1>
    </div>

```python
sns.set_style("darkgrid")
fig,ax = plt.subplots(2,4,figsize = (22,10))
sns.regplot(x = 'temp',y = 'total_count',data = train_sort,marker = '+',color = 'g',x_bins = 10,ax=ax[0,0])
ax[0,0].set_title("Temperature Vs total count")
sns.regplot(x = 'windspeed',y = 'total_count',data = train_sort,marker = '+',color = 'g',ax=ax[0,1],x_bins = 10)
ax[0,1].set_title("Windspeed Vs total count")
sns.regplot(x = 'humidity',y = 'total_count',data = train_sort,marker = '+',color = 'g',ax=ax[0,2],x_bins = 20)
ax[0,2].set_title("humidity Vs total count")
sns.regplot(x = 'atemp',y = 'total_count',data = train_sort,marker = '+',color = 'g',ax=ax[0,3],x_bins = 20)
ax[0,3].set_title("atemp Vs total count")
sns.regplot(x = train_sort['month'].astype("int64"),y = 'total_count',data = train_sort,marker = '+',color = 'g',ax=ax[1,0],x_jitter=0.7,x_bins = 10)
ax[1,0].set_title("month Vs total count")
sns.regplot(x = train_sort['weekday'].astype("int64"),y = 'total_count',data = train_sort,marker = '+',color = 'g',ax=ax[1,1],x_jitter=0.7,x_bins = 7)
ax[1,1].set_title("weekday Vs total count")
sns.regplot(x = train_sort['weather_condition'].astype("int64"),y = 'total_count',data = train_sort,marker = '+',color = 'g',ax=ax[1,2],x_jitter=0.5,x_bins = 4)
ax[1,2].set_title("weather_condition Vs total count")
sns.regplot(x = train_sort['hour'].astype("int64"),y = 'total_count',data = train_sort,marker = '+',color = 'g',ax=ax[1,3],x_bins = 20)
ax[1,3].set_title("hour Vs total count")

```

##  - Inference 
* <h3> Regression plots give us an idea about the linear relationship between the target and the feature variable</h3>
* <h3> Here we have add some noise to the data inorder to plot the categorical and target variable</h3>
* <h3> For the clear visulization we have created bins from the data</h3>
    

```python

train_sort.plot(x = 'total_count',y= 'hour', kind = 'scatter', c = 'is_holiday', colormap = 'viridis',figsize = (15,6))
plt.title("Total count of bike shares Vs holiday")
```

<h1><span class="label label-success">Quantile - Quantile plots for Normality check </span></h1>
    </div>

```python
import scipy
```

```python
cols = ['temp','atemp','humidity','windspeed','total_count']
for i in cols:
     scipy.stats.probplot(train_sort[i], dist="norm", plot=plt)
     plt.title("quantile-quantile plot of "+i)
     plt.show()
```

##  - Inference 
* <h3> The QQ plot is used check for normality in the distribution using visual method</h3>
* <h3> A line of 45 degree is drawn accross the data points if most the points lie on the line it is said to follow normal distribution</h3>
* <h3> The first 3 plots follows normal distribution, anyhow for more accurate results we can perform normal tests such as shapiro wilk test </h3>
