```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split as tts
```

```python
print(tr.columns)

print(te.columns)
```

```python
train, test = tts(tr, test_size= 0.25, random_state= 33)
```

```python
from lazypredict.Supervised import LazyClassifier, LazyRegressor
```

```python
x_train = train.drop(['price_range'], axis=1)
y_train = train['price_range']

x_test = test.drop(['price_range'], axis=1)
y_test = test['price_range']
```

```python
clf = LazyClassifier(predictions=True)
models, predictions = clf.fit(x_train, x_test, y_train, y_test)
```

```python
models
```

```python
sc = StandardScaler()

x_train= sc.fit_transform(x_train)
```

```python
x_test= sc.fit_transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier

dtc = DecisionTreeClassifier()

dtc = dtc.fit(x_train,y_train)
y_pred = dtc.predict(x_test)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier # Import Decision Tree Classifier

rfc = RandomForestClassifier()

rfc = rfc.fit(x_train,y_train)
y_pred = rfc.predict(x_test)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()

lr = lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis()

lda = lda.fit(x_train,y_train)
y_pred = lda.predict(x_test)

from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
