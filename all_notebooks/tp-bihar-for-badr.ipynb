```python
credit_card.drop(credit_card.columns.tolist()[-2:],axis=1,inplace=True)
```

```python
# Requête 2

credit_card.groupby("Gender").agg(Effectif=("Gender","count"),Age_Moyen=("Customer_Age","mean"))
```
