```python
cat = []
num = []
for col in df.columns:
    if df[col].dtype=='O':
        cat.append(col)
    else:
        num.append(col)  
        
        
num 
```

```python
plt.style.use('dark_background')
for col in df[num].drop(['cod_ine'],axis=1):
    plt.figure(figsize=(8,5))
    plt.plot(df[col].value_counts(),color='Orange')
    plt.xlabel(col)
    plt.ylabel('cod_ine')
    plt.tight_layout()
    plt.show()
```
