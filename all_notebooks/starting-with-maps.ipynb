```python
import folium
from folium import Choropleth, Circle, Marker
from folium.plugins import HeatMap, MarkerCluster
```

```python
df = pd.read_csv("../input/geospatial-learn-course-data/crimes-in-boston/crimes-in-boston/crime.csv", encoding='latin-1')
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.sum().isnull()
```

```python

m_2 = folium.Map(location=[42.32,-71.0589])
                 
for i,r in df.iterrows():
    Marker([r['Lat'], r['Long']],popup="Robbery cases",icon=folium.Icon(color='red',prefix='fa',icon='info')).add_to(m_2)


m_2
```

```python

m_3 = folium.Map(location=[42.32,-71.0589],zoom_start=15)
mc = MarkerCluster()
for i, r in df.iterrows():
    mc.add_child(Marker([r['Lat'], r['Long']],icon=folium.Icon(color='red')))
m_3.add_child(mc)
m_3
```
