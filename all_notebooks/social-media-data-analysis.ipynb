## Import the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read the data

```python
df = pd.read_csv('../input/social-media-data-analysis/insight_001.csv')
```

## Explore the data set

```python
# Drop first row
df = df.iloc[1:,:]
```

```python
f'The data has {df.shape[0]} Rows and {df.shape[1]} Columns'
```

```python
df.head()
```

## Lifetime Total Likes column

```python
plt.figure(figsize=(10,6))
plt.plot(df['Lifetime Total Likes'], df['Date'], color='orange')
plt.title('Lifetime Total Likes', size=20)
plt.ylabel('Date', size=15)
plt.xlabel('Number of likes', size=15)
plt.xticks(rotation=60)
plt.show()
```

## Daily new likes column

```python
plt.figure(figsize=(10,6))
df_ = df.copy()
df_ = df_.iloc[:,:3]
df_.dropna(inplace=True)
plt.scatter(df_['Daily New Likes'], df_['Date'], color='r')
plt.title('Daily new likes', size=20)
plt.ylabel('Index in the data frame', size=15)
plt.xlabel('Number of likes', size=15)
plt.show()
```

## Daily Unlikes column

```python
plt.figure(figsize=(10,6))
df_ = df.copy()
df_ = df_.iloc[:,[0,3]]
df_.dropna(inplace=True)
plt.scatter(df_['Daily Unlikes'], df_['Date'], color='purple')
plt.title('Daily unlikes', size=20)
plt.ylabel('Date', size=15)
plt.xlabel('Number of unlikes', size=15)
plt.show()
```

## 28 Days Page Engaged Users

```python
plt.figure(figsize=(10,6))
df_ = df.copy()
df_ = df_.iloc[:,[0,6]]
df_.dropna(inplace=True)
df_ = df_.sort_values('28 Days Page Engaged Users')
plt.plot(df_['28 Days Page Engaged Users'],df_['Date'], color='springgreen')
plt.title('Daily unlikes', size=20)
plt.ylabel('Date', size=15)
plt.xlabel('Number of Daily Engaged users', size=15)
plt.xticks(rotation=60)
plt.show()
```
