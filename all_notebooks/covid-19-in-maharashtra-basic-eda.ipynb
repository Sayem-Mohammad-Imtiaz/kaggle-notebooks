```python
# Loading libraries.
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

plt.style.use('tableau-colorblind10')
```

```python
# Reading data.
data = pd.read_csv("../input/latest-covid19-cases-maharashtra-india/Maharashtra Latest Covid Cases.csv")
data.head()
```

```python
data.dtypes
```

```python
data.shape
```

```python
# Remove space from the column names.

data.rename(columns={col:col.replace(' ', '') for col in data.columns}, inplace=True)
data.head()
```

```python
# Summary.
data.describe()
```

```python
data.info()
```

```python
GT500_df = data.loc[data.ActiveCases > 500]
LT500_df = data.loc[data.ActiveCases <= 500]
```

```python
GT500_df.shape
```

```python
# Examine Maharashtra districts where active covid cases are greater than 500
# Total Positive and Recovered cases by Districts.

df = GT500_df.loc[:, ['Districts', 'PositiveCases', 'Recovered']]
df.set_index('Districts', inplace=True)

df.plot(kind='bar', figsize=(14, 8), color=['#54436B', '#50CB93'])

plt.title("Total Positive & Recovered Cases in Districts of Maharashtra.")

plt.xticks(rotation=90)
plt.legend()

plt.show()
```

```python
# Total Active cases by Districts.
plt.figure(figsize=(14, 8))

sns.barplot(x='Districts', y="ActiveCases", color="red", label="Active", data = GT500_df)

plt.title("Total Active Cases.")

plt.xticks(rotation=90)
plt.legend()

plt.show()
```

More than 10,000 active cases are found in **Pune**, **Kolhapur**, **Thane**, **Sangli** and **Mumbai** 

```python
# Total Deaths by Districts.
plt.figure(figsize=(14, 8))

sns.barplot(x='Districts', y="Deceased", color="#333", label="Deceased", data = GT500_df.sort_values(by='Deceased', ascending=False))

plt.title("Total Number of Deceased.")

plt.xticks(rotation=90)
plt.legend()

plt.show()
```

Number of observation in **Mumbai** and **Pune** is high.

```python
# Examine Maharashtra districts where active covid cases are less than 500
# Total Positive and Recovered cases by Districts.

df = LT500_df.loc[:, ['Districts', 'PositiveCases', 'Recovered']]
df.set_index('Districts', inplace=True)

df.plot(kind='bar', figsize=(14, 8), color=['#54436B', '#50CB93'])

plt.title("Total Positive & Recovered Cases in Districts of Maharashtra.")

plt.xticks(rotation=90)
plt.legend()

plt.show()
```

```python
# Total ActiveCases by Districts.
plt.figure(figsize=(14, 8))

sns.barplot(x='Districts', y="ActiveCases", color="red", label="Active", data = LT500_df)

plt.title("Total active cases.")

plt.xticks(rotation=90)
plt.legend()

plt.show()
```

**Yavtamal** and **Jalna** has less than 50 active cases.

```python
# Total Deaths by Districts.
plt.figure(figsize=(14, 8))

sns.barplot(x='Districts', y="Deceased", color="#3E2C41", label="Deceased", data = LT500_df.sort_values(by="Deceased", ascending=False))

plt.title("Total number of Deaths.")

plt.xticks(rotation=90)
plt.legend()

plt.show()
```

Higher observation found in  **Jalgaon**, **Nanded**, and **Yavtamal**.

```python
# Average recovery and fatality rate in Maharashtra.
avg_recovery_rate = round(data['RecoveryRate(%)'].mean(), 2)
avg_fatality_rate = round(data['FatalityRate(%)'].mean(), 2)
col = ['RecoveryRate', 'FatalityRate']

plt.figure(figsize=(10, 8))
plt.pie([avg_recovery_rate,avg_fatality_rate], labels=col, explode=[0.1, 0], autopct="%0.2f")

plt.title('Average Recovery and Fatality Rates in Maharashtra')
plt.show()
```

Average Recovery rate in Maharashtra is 94% and almost 6% of Fatality Rate.
