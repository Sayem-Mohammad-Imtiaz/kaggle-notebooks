# Exploring Europe's Impact on Climate Change

Using carbon dioxide emission data from different sources to compare the impact of Europe on climate change in comparison to the entire planet.

# Importing the data

```python
import numpy as np 
import pandas as pd 
```

```python
#Loading the emission data for the world
cw = pd.read_csv(r'../input/climate-change-where-we-are-and-what-we-can-do/CO2 emissions by energy source World 1990-2017 - World.csv')
#Creating a data frame with the appropriate headers
ccw = pd.DataFrame(cw, columns = ['Year', 'Coal', 'Oil', 'Natural gas', 'Other'])
ccw.head(7)
```

```python
#Repeating for the Europe data
ce = pd.read_csv(r'../input/climate-change-where-we-are-and-what-we-can-do/CO2 emissions by energy source World 1990-2017 - Europe.csv')
cce = pd.DataFrame(ce, columns = ['Year', 'Coal', 'Oil', 'Natural gas', 'Other'])
cce.head(7)
```

# Line graphs

```python
import plotly.express as px

fig1 = px.line(ccw, x='Year', y=['Coal', 'Oil', 'Natural gas', 'Other'], title='The Planets Emissions by Year and Type', labels={
    "value":"Carbon Emissions",
    "variable":"Source"
})
fig1
```

```python
import plotly.express as px

fig2 = px.line(cce, x='Year', y=['Coal', 'Oil', 'Natural gas', 'Other'], title='Europe Emissions by Year and Type', labels={
    "value":"Carbon Emissions",
    "variable":"Source"
})
fig2
```

From the line graphs above, it seems that the planet's carbon emissions have increased since 1990 from all sources, whereas in Europe there has been a steady decrease in carbon emissions from both oil and coal.

# Comparing the total emission changes

```python
#Finding the total emissions for the planet by year
sum_w = ccw.sum(axis=1)
sum_w
```

```python
#Finding the total emissions for Europe by year
sum_e = cce.sum(axis=1)
sum_e
```

```python
#Percentage change in emissions for the planet
change_w = sum_w[6]/sum_w[0]
change_w
```

```python
#Percentage change in emissions for Europe
change_e = sum_e[6]/sum_e[0]
change_e
```

The total carbon emissions for the planet grew by 55% from 1990 to 2017, whilst the total carbon emissions for Europe dropped by 15%, showing how Europe's efforts to tackle climate change are successful in comparison to the planet as a whole.
