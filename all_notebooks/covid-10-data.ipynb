```python
df = pd.read_csv('/kaggle/input/2020-corona-virus-timeseries/COVID-19_geo_timeseries_ver_0311.csv', parse_dates=['update_time']) #, index_col='update_time')
df.head()
```

```python
df.info()
```

```python
df.drop(['country_flag'], axis=1, inplace=True)
df.head()
```

```python
df.info()
```

```python
df['region'].unique()
```

```python
df['recovered']=df['recovered'].fillna(0)
df['deaths'] = df['deaths'].fillna(0)
```

```python
df.tail(10)
```

```python
df['recoveredN'] = df['confirmed_cases'] - df['deaths']
```

```python
df.head()
```

```python
df.drop(['recovered', 'data_source', 'country_code', 'province'], axis=1, inplace=True)
df.head(5)
```

```python
df.info()
```

```python
df = df.dropna(how='any', subset=['longitude', 'latitude'])
df.info()
```

```python
df['deaths'] = df['deaths'].astype(int)
df['recoveredN'] = df['recoveredN'].astype(int)
```

```python
alt.data_transformers.disable_max_rows()
alt.Chart(df).mark_point(filled=True).encode(
    alt.X('longitude'),
    alt.Y('latitude'),
    alt.Size('confirmed_cases'),
    #alt.Color('deaths'),
    alt.OpacityValue(0.7),
    tooltip = [alt.Tooltip('country'),
               alt.Tooltip('longitude'),
               alt.Tooltip('latitude'),
               alt.Tooltip('confirmed_cases')
              ]
).interactive()
```

```python
slider = alt.binding_range(min=1, max=3, step=1)
month= alt.selection_single(name="month", fields=['month'], bind=slider)
alt.Chart(df).mark_point(filled=True).encode(
    alt.X('longitude'),
    alt.Y('latitude'),
    alt.Size('confirmed_cases'),
    #alt.Color('deaths'),
    alt.OpacityValue(0.7),
    tooltip = [alt.Tooltip('country:N'),
               alt.Tooltip('longitude:Q'),
               alt.Tooltip('latitude:Q'),
               alt.Tooltip('confirmed_cases:Q')
              ]
).add_selection(month).transform_filter(month)
```

```python
slider = alt.binding_range(min=1, max=3, step=1)
month= alt.selection_single(name="month", fields=['month'], bind=slider)
alt.Chart(df).mark_point(filled=True).encode(
    alt.X('longitude'),
    alt.Y('latitude'),
    alt.Size('confirmed_cases'),
    alt.Color('region:N'),
    alt.OpacityValue(0.7),
    tooltip = [alt.Tooltip('country:N'),
               alt.Tooltip('longitude:Q'),
               alt.Tooltip('latitude:Q'),
               alt.Tooltip('confirmed_cases:Q')
              ]
).add_selection(month).transform_filter(month).properties(
    width=800,
    height=400
)
```
