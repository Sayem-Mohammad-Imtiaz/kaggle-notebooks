```python
#Code by Andrea Sindico  https://www.kaggle.com/asindico/love-for-poems/notebook

def removePunctuation(x):
    x = x.lower()
    x = re.sub(r'[^\x00-\x7f]',r' ',x)
    x = x.replace('\r','')
    x = x.replace('\n','')
    x = x.replace('  ','')
    x = x.replace('\'','')
    return re.sub("["+string.punctuation+"]", " ", x)
```

```python
#https://stackoverflow.com/questions/51534586/add-and-remove-words-from-the-nltk-stopwords-list

from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

#add words that aren't in the NLTK stopwords list
new_stopwords = ['and', 'at', 'so', 'er']
new_stopwords_list = stop_words.union(new_stopwords)

#remove words that are in NLTK stopwords list
not_stopwords = {'anything', 'free', 'have'} 
final_stop_words = set([word for word in new_stopwords_list if word not in not_stopwords])

print(final_stop_words)
```

```python
#Code by Andrea Sindico  https://www.kaggle.com/asindico/love-for-poems/notebook

def processText(x):
    x= removePunctuation(x)
    #x= removeStopwords(x)
    return x
```

```python
#Since most similar is now in KeyedVectors I copied the snippet below
#https://github.com/RaRe-Technologies/gensim/blob/develop/gensim/models/keyedvectors.py

from gensim import models
from gensim.models import KeyedVectors


import gensim.downloader as api
word_vectors = api.load("glove-wiki-gigaword-100")  # load pre-trained word-vectors from gensim-data 
# Check the "most similar words", using the default "cosine similarity" measure.
result = word_vectors.most_similar(positive=['woman', 'king'], negative=['man'])

most_similar_key, similarity = result[0]  # look at the first match
print(f"{most_similar_key}: {similarity:.4f}")                                                                                 
```
