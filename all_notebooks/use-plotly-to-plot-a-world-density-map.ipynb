```python
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot, plot
init_notebook_mode(connected=True)
```

```python
import pandas as pd
df = pd.read_csv("../input/country-density/Country Density.csv")
```

```python
#Other Projection type: https://plotly.com/python/map-configuration/
data = dict(
        type = 'choropleth',
        colorscale = 'Viridis',
        reversescale = True,
        locations = df['Country'],
        locationmode = "country names",
        z = df['Density'],
        text = df['Country'],
        colorbar = {'title' : 'World Density'},
      ) 

layout = dict(title = 'World Density Map',
                geo = dict(showframe = False,projection = {'type':'natural earth'})
             )
```

```python
choromap = go.Figure(data = [data],layout = layout)
plot(choromap,validate=False)
```
