```python
from collections import Counter
import matplotlib.pyplot as plt

heart = Counter(data_df["output"])
classes = []
count = []
for key in heart.keys():
    classes.append(key)
    count.append(heart[key])
    
colors = ["red", "gray"]

plt.pie(count,
        labels=classes,
        shadow=True,
        colors=colors,
        startangle=80,
        autopct="%1.2f%%")

plt.axis("equal")
plt.tight_layout()
plt.title("Percentage of person with heart disease attack in the dataset", fontsize=15)
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def print_score(y_pred, y_real):
    print("Accuracy: ", accuracy_score(y_real, y_pred))

    print()
    print("Macro precision_recall_fscore_support (macro) average")
    print(precision_recall_fscore_support(y_real, y_pred, average="macro"))

    print()
    print("Macro precision_recall_fscore_support (micro) average")
    print(precision_recall_fscore_support(y_real, y_pred, average="micro"))

    print()
    print("Macro precision_recall_fscore_support (weighted) average")
    print(precision_recall_fscore_support(y_real, y_pred, average="weighted"))
```

```python
x = data_df.drop("output", axis=1)
y = data_df["output"]
```

```python
from sklearn.model_selection import train_test_split

train_x, validation_x, train_y, validation_y = train_test_split(x, y, test_size=0.1, random_state=1, shuffle=True)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler 
```

```python
pca = PCA(n_components=10)
scaler = MinMaxScaler()

train_x = pca.fit_transform(train_x)
train_x = scaler.fit_transform(train_x)
```

```python
validation_x = pca.transform(validation_x)
validation_x = scaler.transform(validation_x)
```

```python
from sklearn.linear_model import LogisticRegression

%time

param_grid = { "C":[1, 2, 3],
               "penalty":["l1","l2", "elasticnet"],
               "solver": ["newton-cg", "lbfgs", "liblinear"],
               "max_iter": [100, 200, 400]}


grid = GridSearchCV(LogisticRegression(), param_grid, refit=True, verbose=1)
```

```python
grid.fit(train_x, train_y)

print(grid.best_params_)
print(grid.best_estimator_)
```

```python
y_pred = grid.predict(train_x)
print_score(y_pred, train_y)
```

```python
y_pred = grid.predict(validation_x)
print_score(y_pred, validation_y)
```
