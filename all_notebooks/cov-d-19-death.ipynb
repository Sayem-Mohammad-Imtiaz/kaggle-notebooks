```python
countriesList = []
for country in covid_19["Country/Region"]:
    for unique in (country.split(",")):
        if unique not in countriesList: 
            countriesList.append(unique)
print(countriesList)
```

```python
countryTotal = {} #Total counties


for country in countriesList:
    countryTotal[country] = 0
    
print(countryTotal)
```

```python
for idx, countries in enumerate (covid_19["Country/Region"]):
    #print(idx,genres)
    for country in countries.split(","):
        countryTotal[country] += covid_19["Deaths"][idx]
        
print(countryTotal)
```
