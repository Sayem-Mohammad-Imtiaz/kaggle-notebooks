```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../input/list-of-700-unicorn-companies-across-the-globe/List of 700 Unicorn Companies.csv')
```

```python
data['Country of origin'].value_counts()
```

```python

data['Country of origin'].value_counts()[:10].sort_values().plot(kind = 'barh')
```

```python
data['City'].value_counts()
```

```python
data['City'].value_counts()[:10].sort_values().plot(kind = 'barh')
```

```python
data['Industry'].value_counts()
```

```python
data['Industry'].value_counts()[:10].sort_values().plot(kind = 'barh')
```

```python
data['Valuation (in $B)'] = data['Valuation (in $B)'].apply(lambda x : x.replace('$', ''))
```

```python
data['Valuation (in $B)'] = pd.to_numeric(data['Valuation (in $B)'], downcast="float")
```

```python
data['Valuation (in $B)'].describe()
```

```python
type(data)
```

```python
data.groupby(['Country of origin'])['Valuation (in $B)'].sum()
```

```python
data.groupby(['Country of origin'])['Valuation (in $B)'].sum().plot(kind = 'barh')
```

```python
grouped = data.groupby('Country of origin')['Valuation (in $B)'].sum().reset_index()
top_10 = grouped.sort_values('Valuation (in $B)', ascending=False).head(10)
top_10.head(10)
```

```python
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="Valuation (in $B)", y="Country of origin", data=top_10)
```

```python
data.groupby(['Industry'])['Valuation (in $B)'].sum().plot(kind = 'barh')
```

```python
grouped = data.groupby('Industry')['Valuation (in $B)'].sum().reset_index()
top_10 = grouped.sort_values('Valuation (in $B)', ascending=False).head(10)
top_10.head(10)
```

```python
sns.set_theme(style="whitegrid")
ax = sns.barplot(x="Valuation (in $B)", y="Industry", data=top_10)
```
