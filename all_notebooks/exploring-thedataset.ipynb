```python
j = df['Job Title'].value_counts()
```

```python
#to find the percentage of applicants for each job title 

#On a sidenote I'd be really greatful if anyone can make this chart a little bit more appealing to the eye.
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n".format(pct, absolute)




fig, ax = plt.subplots(figsize =(10, 10))
wedges, texts, autotexts = ax.pie(j, 
                                  autopct = lambda pct: func(pct, j), 
                                  labels = j.keys(),
                                  startangle = 90,
                                  textprops = dict(color ="black"))
```
