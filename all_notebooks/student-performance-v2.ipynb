```python
import pandas as pd
import numpy as np
import seaborn as sb
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from tpot import TPOTRegressor
```

```python
df= pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
df
```

```python
df.rename(columns={'race/ethnicity':'race_ethnicity'},inplace=1)
df
```

```python
Gender_Dummy = pd.get_dummies(df['gender'])
Race_Ethnicity_Dummy = pd.get_dummies(df['race_ethnicity'])
Parental_Ed_Dummy= pd.get_dummies(df['parental level of education'])
Lunch_Dummy = pd.get_dummies(df['lunch'])
Test_Pre_Dummy = pd.get_dummies(df['test preparation course'])
```

```python
# Merge all the endoded to the DF
df = pd.concat((df,Gender_Dummy,Race_Ethnicity_Dummy,Parental_Ed_Dummy,Lunch_Dummy,Test_Pre_Dummy), axis=1)
df.rename(columns={'free/reduced':'free_reduced'},inplace=1)
df
```

```python
df.info()
```

```python
X=df.drop(['gender','race_ethnicity','parental level of education','lunch','test preparation course','math score','reading score','writing score','female','male'], axis=1)
```

```python
X
```

```python
X.info()
```

```python
y=df['math score']
```

# Data Anaysis using TPOT

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X, y, train_size=0.80, test_size=0.20)
```

```python
y_train1
```

```python
X_train1
```

```python
X_train=np.array(X_train1)
y_train=np.array(y_train1)
X_test=np.array(X_test1)
y_test=np.array(y_test1)
```

```python
X_train.shape
```

```python
pipeline_optimizer = TPOTRegressor()


#tpot = TPOTRegressor(verbosity=3, max_time_mins=2, max_eval_time_mins=0.04, population_size=40)

pipeline_optimizer = TPOTRegressor(max_time_mins=10,generations=10, population_size=20, cv=5,
                                    random_state=42, verbosity=2)
pipeline_optimizer.fit(X_train,y_train)
```

```python
print(pipeline_optimizer.score(X_test, y_test))
```

```python
pipeline_optimizer.export('tpot_exported_pipeline.py')
```
