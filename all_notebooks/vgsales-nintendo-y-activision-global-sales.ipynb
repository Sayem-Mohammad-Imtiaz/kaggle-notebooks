# Graficando algunos datos

```python
col=['Publisher','Year','Global_Sales']
df=data.loc[(data.Publisher.isin(['Activision','Nintendo'])) & (data.Year>=2006),col]
df.head()
```

```python
df_1=df.groupby(['Publisher','Year']).Global_Sales.sum()
print(df_1)
type(df_1)
```

```python
df_1.plot(kind='bar')
```

```python
df_2=df_1.reset_index()
#print(df_2)
df_3=pd.DataFrame(df_2)
df_3.head()
```

```python
plt.figure(figsize=(12,6))
plt.title('Ventas acumuladas por a√±o')
sns.barplot(x="Year", y="Global_Sales", hue="Publisher", data=df_3)
plt.ylabel('Ventas en millones')
plt.show()
```
