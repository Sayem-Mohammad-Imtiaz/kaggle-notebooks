<h2 style='border:3px solid DodgerRed;background-color:#9d0aea;font-family:comic sans ms'><center>Detailed understanding of student scores</center></h2>

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import plotly.figure_factory as ff
```

```python
df = pd.read_csv(r"../input/student-study-hour-v2/Student Study Hour V2.csv")
```

```python
df
```

<h2 style='border:3px solid DodgerRed;background-color:#9d0aea;font-family:comic sans ms'><center>How many hours he had studied to get below 35 marks</center></h2>

<h1 style = "font-family: Comic Sans MS;color:Green">Lets try to make this dataset to be more meaningful so that things will be very clear</h1>

<h3 style = "font-family: Comic Sans MS;color:Red">if hours can be converted into minutes then it will be more meaningfull</h3>

```python
df["Hours"] = df["Hours"]*60
```

```python
df["Hours"] = df["Hours"].astype(int)
```

```python
fail = df[df["Scores"]<35].head(11) 
fail.sort_values(by="Scores")
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(data=fail,x="Hours",y="Scores")
```

<h2 style='border:3px solid DodgerRed;background-color:#9d0aea;font-family:comic sans ms'><center>How many hours he had studied to get above 90 marks</center></h2>

```python
success = df[df["Scores"]>=90]
```

```python
ff.create_table(success)
```

<h2 style='border:3px solid DodgerRed;background-color:#9d0aea;font-family:comic sans ms'><center>How many hours students are studied to get an above average marks</center></h2>

```python
average = df[(df["Scores"]<90) & (df["Scores"]>60)] 
ff.create_table(average)
```

<h2 style='border:3px solid DodgerRed;background-color:#9d0aea;font-family:comic sans ms'><center>How many hours students are studied to get an below average marks</center></h2>

```python
below_average = df[(df["Scores"]<60) & (df["Scores"]>35)] 
```

```python
ff.create_table(below_average)
```
