# Drug Classification - Decision Trees

## Importing libraries

```python
import numpy as np 
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

## Importing the dataset

```python
df = pd.read_csv("../input/drugs-a-b-c-x-y-for-decision-trees/drug200.csv")
df[0:10]
```

## Data pre-processing

```python
x = df[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']].values
```

```python
from sklearn import preprocessing
sex = preprocessing.LabelEncoder()
sex.fit(['F','M'])
x[:,1] = sex.transform(x[:,1]) 


BP = preprocessing.LabelEncoder()
BP.fit([ 'LOW', 'NORMAL', 'HIGH'])
x[:,2] = BP.transform(x[:,2])


Chol = preprocessing.LabelEncoder()
Chol.fit([ 'NORMAL', 'HIGH'])
x[:,3] = Chol.transform(x[:,3]) 

x[0:10]
```

```python
y = df["Drug"]
y[0:10]
```

## Setting up decision tree

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=3)
```

## Modelling

```python
dt = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
dt
```

```python
dt.fit(x_train,y_train)
```

## Prediction

```python
pt = dt.predict(x_test)
print(pt[0:5])
print(y_test[0:5])
```

## Evaluation

```python
from sklearn import metrics
print("Accuracy of the decision tree: ", metrics.accuracy_score(y_test,pt))
```
