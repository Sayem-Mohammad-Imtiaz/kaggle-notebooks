```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import math
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr

```

```python
df= pd.read_csv("../input/linear-regression-dataset/Linear Regression - Sheet1.csv")
print(df.head(5))
```

We start by checking if there are any duplicates, NaN or null values.

```python
print(df.pivot_table(index = df, aggfunc ='size'))
print(df.isna().sum())
print(df.isnull().sum())
```

#Calculation of the correlation matrix between the two variables to see if there exist any relationship, then the performance of the pearson T_Test.
##Based on the obtained results, there is a very high positive correlation between the X and the Y variables.
###Based on the obtained p-value and the T_test there is enough evidence to reject the H0 hypothesis in favor of H1.
####This means that there stronly exist a positive linear relationship between the data_set variables.

```python
matrix_corr=df.corr()
print(matrix_corr)
Pearson_Test= pearsonr(df["X"],df["Y"])
print("Pearson_Test rsult=",Pearson_Test)
```

Graphical representation of the data_set in a scatterplot

```python
Y=pd.DataFrame(df['Y'])
X=pd.DataFrame(df['X'])
plt.scatter(Y,X, linewidths=0.5, edgecolors='r', marker='o')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

Linear Regression analysis with the "statsmodels library"

```python
RL=sm.OLS(X,Y)
RL=RL.fit()
print(RL.summary())
```

The correlation of determination (R-squared) is 0.98, meaning that 98% of the variance in the variable Y is
due to the variance in the variable X.


Linear Regression analysis with the "sklearn.linear_model" library

```python
reg=LinearRegression()
reg=reg.fit(X,Y)
reg=LinearRegression().fit(X,Y)
print("Slope=",reg.coef_)
print("Intercept=",reg.intercept_)
Y_predict=reg.predict(X)
```

Estimated regression equation:
y=5.8887997+0.64004489*X

Distplot of the Y Real Values vs Y Predicted Values

```python
sns.distplot(Y, hist=False, label='Real Values')
plt.legend("Real Values")
sns.distplot(Y_predict, hist=False, label='Predicted Values')
plt.legend("Predicted Values")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Real Values vs Predicted Values")
plt.show()
```

Residuals Plot

```python
sns.residplot(X,Y)
plt.show()
```

Measure of the SSE


```python
SSE=Y-Y_predict
SSE=SSE**2
print("SSE=",SSE.sum().round())
```

Measure of the SSR

```python
mean=Y.mean()
print(mean)
Y_mean=np.ones((300,1),dtype=np.int16)
Y_mean=Y_mean*102.215556
SSR=Y_predict-Y_mean
SSR=SSR**2
print("SSR=",SSR.sum().round())
```

Measure of SSTO

```python
SSTO=SSE+SSR
print("SSTO=",SSTO.sum().round())
```

```python
#df=n-2=300-2=298
MSE=SSE/298
print("MSE=",MSE.sum().round())
RMSE=math.sqrt(264)
print("RMSE=",RMSE)
```
