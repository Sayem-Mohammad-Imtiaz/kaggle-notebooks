```python
import pandas as pd # package for high-performance, easy-to-use data structures and data analysis
import numpy as np # fundamental package for scientific computing with Python
import gc
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load and merge datasets. It may take a little while as the datasets are large.
df= pd.read_csv("../input/netflix-shows/netflix_titles.csv")
df.shape
df.head()
```

```python
df.dtypes
```

```python
df.listed_in
    
```

```python
#select columns
df = df.filter(["show_id", "type", "country", "listed_in"])
df.dropna()
df[:20]
```

```python
df[:15]
```

*Japan*

```python
#df for japan only
df_japan = df.filter(["show_id", "type", "country", "listed_in"])

```

```python

df_japan['japan'] = df_japan["country"].str.contains("Japan")
indexNames = df[df_japan['japan'] == False].index
df_japan.drop(indexNames, inplace=True)
df_japan[:10]
```

```python
df_japan.dropna()
```

```python
df_japan.columns
#there's no japan column added it's just a function that checks values
```

```python
df_japan["listed_in"].value_counts().head(20)
```

```python
#contents available in japan as a ppercent
df_japan["listed_in"].value_counts().plot(kind = "pie" ,figsize=(20,20),autopct='%1.1f%%')
```

```python

df_japan["romantic"] = df_japan["listed_in"].str.contains("Romantic")
df_japan["romantic"].value_counts().plot.bar(figsize = (12,8))
```

```python
df_japan["romantic"].value_counts().plot(kind = "pie" ,figsize=(8,8),autopct='%1.1f%%')
```

*India*

```python
#df for india only
df_india = df.filter(["show_id", "type", "country", "listed_in"])
df_india['india'] = df_india["country"].str.contains("India")
indexNames = df[df_india['india'] == False].index
df_india.drop(indexNames, inplace=True)
df_india[:10]
```

```python
df_india.dropna()
```

```python
df_india["listed_in"].value_counts().head(20)
```

```python
#contents available in india as a ppercent
df_india["listed_in"].value_counts().plot(kind = "pie" ,figsize=(20,20),autopct='%1.1f%%')
```

```python
df_india["romantic"] = df_india["listed_in"].str.contains("Romantic")
df_india["romantic"].value_counts().plot.bar(figsize = (12,8))
```

```python
# romantic programmes available in india
df_india["romantic"].value_counts().plot(kind = "pie" ,figsize=(8,8),autopct='%1.1f%%')
```

# **Calculation**

* Japan Romatic programs available = 11.1%
* India Romantic programs available = 12.4%
* p-value = 0.276378
>** Meaning that, there's no significant difference.**
