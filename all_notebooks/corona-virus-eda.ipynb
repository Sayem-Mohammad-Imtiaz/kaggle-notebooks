```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
import random
import matplotlib.colors as mcolors
```

```python
data = pd.read_csv("../input/clean-cov/cronovirus_clean_upddated.csv")
data.head()
```



    Province/State - City of virus suspected cases.
    Country/Region - Country of virus suspected cases.
    last update - Date of update of patient infected
    Confirmed - Confirmation by doctors that this patient is infected with deadly virus
    Suspected - Number of cases registered
    Recovered - Recovery of the patient
    Deaths - Death of the patient .

### Confirm Cases againt Provinces

```python
unique_provinces = data['Province/State'][data.Confirmed >0 ].unique()
```

```python
province_confirmed_cases = []
for i in unique_provinces:
    province_confirmed_cases.append(data[data.Confirmed > 0 ][data['Province/State'] == i].Confirmed.sum())
```

```python
for i in range(len(unique_provinces)):
    print (f'{unique_provinces[i]}: {province_confirmed_cases[i]} cases')
```

```python
c = random.choices(list(mcolors.CSS4_COLORS.values()),k = len(unique_provinces))
plt.figure(figsize=(15,15))
plt.pie(province_confirmed_cases, colors = c)
plt.legend(unique_provinces, loc='best')
plt.show()
```

### Separating Date

```python
import datetime 

data['Last Update'] = pd.to_datetime(data['Last Update']) 
data['Date'] = [datetime.datetime.date(d) for d in data ['Last Update']]
data['Time'] = [datetime.datetime.time(d) for d in data['Last Update']]
```

```python
data['Date'] = data['Date'].astype(str)
day = data["Date"].values
day = [my_str.split("-")[2] for my_str in day]
data["Date"] = day
```

```python
data = data.drop(['Last Update'], axis=1)
```

```python
data.head()
```

```python
unique_dates = data['Date'].unique()
unique_dates = np.flipud(unique_dates) 

unique_dates
```

### Confirm cases 

```python
china_cases = []

for i in unique_dates:
    china_cases.append(data[data['Date']==i].Confirmed.sum())

plt.figure(figsize=(15, 10));
plt.plot(unique_dates, china_cases);
plt.title('No of Coronavirus Cases Over Days', size=15);
plt.xlabel('Days', size=15)
plt.ylabel('No of Cases', size=15);
plt.show();
```

### No of Deaths

```python
Death = data['Death'] > 0
```

```python
china_cases = []

for i in unique_dates:
    china_cases.append(data[data['Date']==i].Death.sum())

plt.figure(figsize=(15, 10));
plt.plot(unique_dates, china_cases);
plt.title('Death cases', size=15);
plt.xlabel('Days', size=15)
plt.ylabel('No of Deaths', size=15);
plt.show();
```

### Deaths in different provinces

```python
province_Death_cases = []
for i in unique_provinces:
    province_Death_cases.append(data[data.Death>0][data['Province/State']==i].Death.sum())
```

```python
c = random.sample(list(mcolors.CSS4_COLORS.values()),k = len(unique_provinces))
plt.figure(figsize=(15,15))
plt.pie(province_Death_cases, colors = c)
plt.legend(unique_provinces, loc='best')
plt.show()
```
