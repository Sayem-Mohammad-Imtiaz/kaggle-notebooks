```python
import matplotlib.pyplot as plt
import seaborn as sns 
from subprocess import check_output
from wordcloud import WordCloud, STOPWORDS
```

1. Loading the data 
2. Number of episodes in each season
3. Director ranked according to most votes
4. Most popular episode in each season(acc to the votes)

# Loading the data

```python
data = pd.read_csv('../input/friends-series-dataset/friends_episodes_v2.csv')
data.head()
```

# Number of episodes in each season

```python
data.groupby('Season')['Episode_Title'].count()
```

# Director ranked according to most votes

```python
data.groupby('Director')[['Stars', 'Votes']].agg(np.mean).sort_values(['Votes'],ascending=False)
```

# Most popular episode in each season(acc to the votes)

```python
data.groupby('Season').max()
```

# Word Cloud (Most used words in the summary) 

```python
stopwords = set(STOPWORDS)
cast = {"Rachel","Ross","Monica","Phoebe","Joey","Chandler"}
stopwords.update(cast)
wordcloud = WordCloud(
                          background_color='white',
                          stopwords=stopwords,
                          max_words=200,
                          max_font_size=40, 
                          random_state=42
                         ).generate("".join(data['Summary']))

print(wordcloud)
fig = plt.figure(1)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

# Director vs Votes

```python
fig = plt.figure(num=None, figsize=(10, 10), dpi=80, facecolor='w', edgecolor='k')
b = sns.stripplot(x="Votes", y="Director", data=data, jitter=True, linewidth=1)
b.set_yticklabels(b.get_ymajorticklabels(), fontsize = 18)
plt.show()
```

# Relation between stars and votes

```python
fig = plt.figure(num=None, figsize=(8, 8), dpi=80, facecolor='w', edgecolor='k')
sns.scatterplot(x="Stars", y="Votes", data=data)
plt.show()
```
