**The below chart shows which king has defended in maximum number of battles and from the chart it can be concluded that *Robb Stark* has defended in maximum number of battles**

**The blow box plot shows the relation between the attacker size and the result of the battles with respect to attacker (i.e. Whether attacker won the battle or lost it)**

```python
sns.boxplot(x='attacker_outcome', y='attacker_size', data=battlesData)
```

**The below box plot shows the relationship between the defender size and the result of the battles with respect to attacker(i.e. Whether the attacker won the battle or lost it)**

```python
# Add new column "defender_outcome" into the data for the below chart
def populate_defender_outcome(attacker_outcome):
    defender_outcome = []
    
    for outcome in attacker_outcome:
        if outcome == 'win':
            defender_outcome.append('loss')
        elif outcome == 'loss':
            defender_outcome.append('win')    
        else:
            defender_outcome.append(np.nan)
    return defender_outcome

defender_outcome = populate_defender_outcome(battlesData.attacker_outcome)

battlesData['defender_outcome'] = defender_outcome
```

```python
sns.boxplot(x='defender_outcome', y='defender_size', data=battlesData)
```

**The below chart shows how the number of major deaths affected the result of battle for the attacker**

```python
sns.barplot(x='attacker_outcome', y='major_death', data=battlesData)
```

**The below chart shows how the number of major capture affacted the result of battle for the attacker**

```python
sns.barplot(x='attacker_outcome', y='major_capture', data=battlesData)
```
