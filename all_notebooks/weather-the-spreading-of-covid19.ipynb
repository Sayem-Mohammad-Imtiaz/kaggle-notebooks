## Relation between Confirmed Cases, Temperature, Humidity and Wind

```python
# we take only these 4 features
sth = monthly_milano[['Time', 'Temperature (Â° F)','Humidity (%)','Wind Speed (mph)']]\
[monthly_milano.Time >= '2020-02-24']
sth= sth.reset_index()

cmi = milan['TotalPositiveCases'].copy()
cmi = cmi.reset_index()
```

```python
cthw = sth[['Time', 'Temperature (Â° F)','Humidity (%)','Wind Speed (mph)']].copy()
cthw['TotalPositiveCases'] = cmi['TotalPositiveCases'].copy()
```

```python
cthw['Time'] = cthw.Time.apply(lambda x: x.strftime('%Y-%m-%d'))
```

```python
cthw.head()
```

```python
# some statistics
cthw.describe()
```

```python
cthw.corr().style.background_gradient('viridis')
```

wind speed and humidity correlation = -0.351 no more correlate
humidity and totalpositive correlation = -0.29

```python
size = cthw['Temperature (Â° F)'].pow(0.4)
fx = px.scatter(cthw, x='Humidity (%)', y='Wind Speed (mph)', hover_name='Time',
          title='Relation between Humidity-Wind Speed in Milano', color='TotalPositiveCases',
          size='TotalPositiveCases')
fx.show()
```

**We can see that despite the small correlation that exists between these variables we find that the number of affected cases increases per day while humidity decreases**.

```python
#siz = cthw.TotalPositiveCases.pow(0.4)
fy = px.scatter(cthw, x='Humidity (%)', y='TotalPositiveCases', hover_name='Time',
          title='Relation between Humidity-TotalPositiveCases in Milano', color='Wind Speed (mph)',
          size = 'TotalPositiveCases')
fy.show()
```

**As the number of positive cases increases and the humidity decreases; we can see very well by day the wind speed gradually increases with the number of positive cases**.

```python
fp = px.scatter(cthw, y='Wind Speed (mph)', x='TotalPositiveCases', hover_name='Time',
          title='Relation between Wind Speed-TotalPositiveCases in Milano', color='Humidity (%)',
          size = 'TotalPositiveCases')
fp.update(layout_coloraxis_showscale=False)
fp.show()
```

**wind speed and totalpositivecases have a nonlinear shape**
