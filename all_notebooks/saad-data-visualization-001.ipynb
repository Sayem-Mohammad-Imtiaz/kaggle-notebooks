```python
df = pd.read_csv("/kaggle/input/zinc-prices/zinc_prices_IMF.csv")
df.head(10)
```

```python
print(df.info())
```

```python
print(df['Date'].count())
print(df['Price'].count())
```

```python
#the date column is of type object change it to float()
df['Date'].unique()
```

```python

# Import label encoder
from sklearn import preprocessing
 
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
 
# Encode labels in column 'species'.
df['Date']= label_encoder.fit_transform(df['Date'])

df["Date"].unique()
```

```python
print(df["Price"].max())
print(df["Price"].min())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(data = df )
plt.show()
```

```python
sns.lineplot(data = df, x = "Date", y = "Price")
plt.show()
```

```python
sns.scatterplot(data = df , x = "Date", y = "Price", hue = "Price")
plt.show()
```

```python
sns.regplot(x = "Date", y = 'Price', 
            # Set scatter point opacity & color
            scatter_kws = {'alpha':0.5, 'color':'gray'}, 
            # Disable confidence band
            ci= False, 
            data = df.loc[:100])

plt.show()
```

```python
fig, ax = plt.subplots()
# Plot the distribution of data
sns.distplot(df['Price'], ax=ax)
# Create a more descriptive x axis label
ax.set(xlabel="Price", xlim=(0,5000), title="Density of Price")
# Show the plot
plt.show()
```
