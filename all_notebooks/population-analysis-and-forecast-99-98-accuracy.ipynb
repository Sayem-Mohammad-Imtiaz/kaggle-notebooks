# Population Analysis

Reading the dataset

```python
pop = pd.read_csv('../input/world-population-by-year/WorldPopulation.csv', header=0, sep=',')
print(pop.head(5))
```

```python
print(pop.info())
```

Analysing change in all variables with respect to the Year,

```python
plt.figure(figsize=(15, 15))
x = pop['Year']
y = pop['Population']
plt.subplot(2,3,1)
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('Population')

x = pop['Year']
y = pop['ChangePerc']
plt.subplot(2,3,2)
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('ChangePerc')

x = pop['Year']
y = pop['NetChange']
plt.subplot(2,3,3)
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('NetChange')

x = pop['Year']
y = pop['Density']
plt.subplot(2,3,4)
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('Density')

x = pop['Year']
y = pop['Urban']
plt.subplot(2,3,5)
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('Urban')

x = pop['Year']
y = pop['UrbanPerc']
plt.subplot(2,3,6)
plt.plot(x,y)
plt.xlabel('Year')
plt.ylabel('UrbanPerc')
```

# Population Prediction

Finding relation between Year and Population:

```python
x = pop['Year']
y = pop['Population']
predict_pop = np.poly1d(np.polyfit(x, y, 3))
print(r2_score(y, predict_pop(x)))
```

**This results in *99.98%* accuracy**

```python
plt.figure(figsize=(10, 10))
x = predict_pop(pop['Year'])
y = pop['Population']
plt.scatter(x,y, alpha=0.7)
plt.xlabel('Predicted population')
plt.ylabel('Actual population')
```
