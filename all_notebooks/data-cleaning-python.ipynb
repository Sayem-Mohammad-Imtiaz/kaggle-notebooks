```python
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#importing data and understanding columns
df = pd.read_csv('../input/loan-dataset/loan.csv')
df.head()
```

```python
df.shape
```

```python
df.dtypes
```

```python
# Display columns missing values %
round(100*(df.isnull().sum()/len(df.index)),2)
```

```python
#Data Cleaning 
print(df.id.nunique())
print(df.member_id.nunique())
#Both columns have same number of unique values therefore Droping one of the columns
df = df.drop(['member_id'], axis=1)
```

```python
#many columns have 100% null values so dropping them 
df = df.dropna(how='all', axis=1)
```

```python
#Droping columns - mths_since_last_delinq,mths_since_last_record,next_pymnt_d 
#having more than 30% nans
df = df.drop(['mths_since_last_delinq', 'mths_since_last_record', 'next_pymnt_d',], axis=1)
```

```python
#droping all columns with only value 0 i.e. acc_now_delinq column, orresponding delinq_amnt column, chargeoff_within_12_mths column,  
df = df.loc[:, (df != 0).any(axis=0)]
```

```python
# Drop pymnt_plan column; only value in that column is n
df = df.drop(['pymnt_plan'], axis=1)
```

```python
#Drop collections_12_mths_ex_med column; only values in that column are 0 and NA
df = df.drop(['collections_12_mths_ex_med'], axis=1)
```

```python
# Drop policy_code column; only value in that column is 1
df = df.drop(['policy_code'], axis=1)
```

```python
# Drop application_type column; only value in that column is INDIVIDUAL
df = df.drop(['application_type'], axis=1)
```

```python
# Drop tax_liens column; only values in that column are 0 and NA
df = df.drop(['tax_liens'], axis=1)
```

```python
# Drop initial_list_status column; only value in that column is F
df = df.drop(['initial_list_status'], axis=1)
```

```python
# Droping url column; it does not hold significance in analysis 
df = df.drop(['url'], axis=1)

# Droping desc column; desc and purpose are almost same
df = df.drop(['desc'], axis=1)

# Droping title and emp_title column; Instead of emp_title salary provides better analysis
df = df.drop(['emp_title','title'], axis=1)
```

```python
df.to_csv('Loan_cleaned1.csv')
```
