```python
import torch
from torch import autograd
```

```python
from torch.utils.data import TensorDataset, DataLoader
```

```python
import pandas as pd
```

```python
train = pd.read_csv('/kaggle/input/boston-house-prices/housing.csv', sep='\s+', header = None, names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'PRICE'])
```

```python
train.drop('B', axis = 1)
```

```python
train_tensor = torch.tensor(train.values)
```

```python
float_tensor = torch.tensor(train_tensor, dtype=torch.float32)
#train_tensor.dtypes.cast(train_tensor, tf.float32)
```

```python
float_tensor.dtype
```

```python
train_features = float_tensor[:,:13]
```

for i in range(0, train_features.shape[0]):
    train_features[i][11] = train_features[i][11] / 100

```python
train_features
```

```python
train_labels = float_tensor[:, 13]
train_labels.dtype
```

```python
train_labels.reshape((506, -1))
```

```python
train_labels.shape
```

```python
batch_size = 10
dataset = TensorDataset(train_features, train_labels)
data_iter = DataLoader(dataset, batch_size, shuffle = True)

for X, y in data_iter:
    print(X, y)
    print(X.dtype)
    print(X.shape)
    print(y.shape)
    break
```

```python
model = torch.nn.Sequential(torch.nn.Linear(13, 1))
```

```python
model[0].weight.data
```

```python
loss = torch.nn.MSELoss(reduction='mean')
```

```python
trainer = torch.optim.SGD(model.parameters(), lr = 0.001)
```

```python
num_epochs = 100
for epoch in range(1, num_epochs + 1):
    for X, y in data_iter:
        trainer.zero_grad()
        #print(X)
        #print(X.shape)
        #print(X.dtype)
        #print(X != X)
        #print(model.forward(X).reshape(-1))
        #print(y)
        l = loss(model.forward(X).reshape(-1), y)
        l.backward()
        trainer.step()
    l = loss(model.forward(train_features).reshape(-1), train_labels)
    if epoch % 5 == 0:
        print('epoch %d, loss: %f' % (epoch, l.item()) , '|\tw', model[0].weight.data, '|\tb', model[0].bias.data)
```
