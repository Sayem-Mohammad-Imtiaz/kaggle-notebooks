```python
import plotly as plt
import plotly.graph_objects as go #Visualization
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns




# Get Data: 
data = pd.read_csv('/kaggle/input/top100realestatemay2021/2021_4_27_Top100RealEstate.csv', encoding="ISO-8859-1")


data.describe()

start,end = 0,100 

fig = go.Figure(data=go.Scatter3d(
    x=data['Price To Income Ratio'][start:end],
    y=data['Affordability Index'][start:end],
    z=data['Mortgage As A Percentage Of Income'][start:end],
    text=data['Country'][start:end],
    mode='markers',
    marker=dict(
        sizemode='diameter',
        sizeref=2.5,
        size=data['Mortgage As A Percentage Of Income'][start:end],
        color = data['Mortgage As A Percentage Of Income'][start:end],
        colorscale = 'Viridis',
        colorbar_title = 'Mortgage As A % Of Income<br>',
        line_color='rgb(140, 140, 170)'
    )
))








fig.update_layout(height=600, width=600,
                  title='Real Estate Market by Country May 2021 - The Top 100')






fig.show()


#describe data
data.describe()

#Sort data
data.sort_values(by='Affordability Index', ascending=True).tail(10)
```
