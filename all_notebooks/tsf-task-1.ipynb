### Task 1 : Predict the percentage of an student based on the number of study hours.

Only two variables are involved , one is Hours which is Independent Variable and other one is Scores which is Dependent Variable. Using Hourse, we have to predict the students score.
#### For this prediction, Linear Regression would be the best choice.

## Importing libraries and collecting data

```python
#import all required libraries and functions
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import file
link="../input/tsf-ds-ba-task-1/st_record.csv"
dataset=pd.read_csv(link)
dataset.head()
```

## Data Preparation

```python
# iloc function : divide the data into features and target
X = dataset.iloc[:,:-1].values  
y = dataset.iloc[:,1].values
```

```python
dataset.info(verbose=True)
```

> ##### Data filled completely

```python
#plot hours vs score
plt.scatter(X,y,c="r",marker="s",label="score")
plt.title('Hours Vs Scores')
plt.xlabel('hours')
plt.ylabel('marks')
plt.legend()
plt.show()
```

#### Split data into Training and Testing

```python
# train_test_split : split the data for training and testing purpose ,0]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
```

#### Training the data

```python
regression=LinearRegression()
regression.fit(X_train,y_train)
```

```python
# coefficient and Intercept
coef = regression.coef_
intercept = regression.intercept_
```

```python
#plot of regression line
line = coef * X + intercept
# Plotting for the test data
plt.scatter(X, y,c='r')
plt.plot(X, line,'-g');
plt.show()
```

#### Testing the data

```python
# prediction of scores based on training
y_pred=regression.predict(X_test)

prediction = pd.DataFrame()
prediction['Hours'] = pd.Series(X_test.flatten())
prediction['Original Scores'] = y_test
prediction['Predicted Scores'] = y_pred
# concat([X_test,y_test,y_pred])
# prediction.columns = ['Hours','Original Scores','Predicted Scores']
prediction.head()
```

#### Accuracy Estimation

```python
print('Mean Absolute Error:',metrics.mean_absolute_error(y_test, y_pred))
```

```python
print('Coefficient of determination :',metrics.r2_score(y_test,y_pred))
```

#### Predict the score for a student studied for 9.25 hrs/day

```python
student_score = coef * 9.25 + intercept
print("Student studying for 9.25 hrs/day may score {:.2f}".format(*student_score))
```

#### Using linear regression, successfully predicted the score of student studying for 9.25 hrs & would score 92.92 marks
