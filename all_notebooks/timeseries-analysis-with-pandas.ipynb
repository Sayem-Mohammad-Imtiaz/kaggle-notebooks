**Numpy can handle datetime objects more efficiently**

```python
import numpy as np
```

```python
np.array(['2021-04-02', '2021-03-09', '2022-01-10'], dtype = 'datetime64')
```

That extra-letter [D] with datetime64 decides the precision, by default it is set to day, if you set it to month then it will consider the datetime upto month only.

```python
np.array(['2021-04-02', '2021-03-09', '2022-01-10'], dtype = 'datetime64[M]')
```

By default if follows the format of Year-Month-Day, it can't handle every kind of date format

```python
np.array(['2021-04-02', '2021-03-09', '2022-01-10'], dtype = 'datetime64[h]') # will consider upto hour
```

**np.arange(start, stop, step)**

This can be applied on datetime objects also but explicitly dtype should be mentioned as step size can either be day, year , month or hour

```python
np.arange('2020-03-01', '2020-03-30', 7, dtype = 'datetime64[D]')
```

```python
np.arange('1960', '2000', 7, dtype = 'datetime64[Y]')
```

# Time Resampling

It is similar to groupby for timer series data

```python
df.resample(rule = 'A').mean()
```

**It gives yearly mean of Close and Volume**

Other kind of rules are here:

B -->        business day frequency<br>
C  -->        custom business day frequency (experimental)<br>
D     -->     calendar day frequency<br>
W    -->      weekly frequency<br>
M     -->     month end frequency<br>
SM     -->    semi-month end frequency (15th and end of month)<br>
BM     -->    business month end frequency<br>
CBM   -->     custom business month end frequency<br>
MS    -->     month start frequency<br>
SMS   -->     semi-month start frequency (1st and 15th)<br>
BMS      -->   business month start frequency<br>
CBMS    -->    custom business month start frequency<br>
Q       -->    quarter end frequency<br>
BQ      -->    business quarter endfrequency<br>
QS      -->    quarter start frequency<br>
BQS     -->    business quarter start frequency<br>
A       -->    year end frequency<br>
BA, BY    -->  business year end frequency<br>
AS, YS    -->  year start frequency<br>
BAS, BYS  -->  business year start frequency<br>
BH    -->     business hour frequency<br>
H     -->     hourly frequency<br>
T, min   -->  minutely frequency<br>
S     -->     secondly frequency<br>
L, ms  -->    milliseconds<br>
U, us  -->    microseconds<br>
N      -->    nanoseconds<br>

We can also create our own custom functions for resampling and apply that using .apply method

```python
df['Close'].resample('A').mean().plot(kind = 'bar', title = 'Yearly mean closing price for Starbucks')
```

```python
title = 'Montly Max Closing price for starbucks'

df['Close'].resample('M').max().plot(kind = 'bar', title = title, figsize = (16,6))
```

## Rolling and Expanding

```python
df['Close'].plot(figsize = (16, 5))
```

**Rolling**

```python
df.rolling(window = 7).mean().head(10) # It is taking the average of every 7 day due to which first 6 rows has null values and the other rows have the average of every previous 7 days including that row itsef
```

```python
df['Close'].plot(figsize = (16, 5))
df.rolling(window = 60).mean()['Close'].plot() 
```

To add legends to the above plot, add one more column to the dataframe of the rooling method

**Expanding**

```python
df['Close'].expanding().mean().plot(figsize = (12, 5))
```

At each datapoint, it contains the mean value of all the previous rows i.e, the last point is the mean value of all the previous rows
