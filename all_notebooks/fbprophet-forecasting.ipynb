```python
from fbprophet import Prophet
from fbprophet.plot import plot_plotly
import pandas as pd
```

```python
def load_data(data_dir):
    df = pd.read_csv(data_dir)
    cols = df.columns
    return df, cols[1:]

def train_model(data, currency, period):
    df_train = data[["date(y-m-d)", currency]]
    df_train = df_train.iloc[-365*2 :]
    df_train = df_train.rename(columns={"date(y-m-d)": "ds", currency: "y"})

    m = Prophet()
    m.fit(df_train)
    future = m.make_future_dataframe(periods=period)
    forecast = m.predict(future)

    return forecast, m
```

```python
data_dir = "/kaggle/input/daily-updated-forex-data-since-201111/forex_usd_data.csv"
data, currencies = load_data(data_dir)
```

```python
currencies
```

```python
currency = currencies[-4]
n_weeks = 4
pred, model = train_model(data, currency, period=n_weeks * 7)
```

```python
fig1 = plot_plotly(model, pred)
fig1.update_layout(xaxis_title="Date", yaxis_title=currency,)
fig1.show()
```
