```python
import pandas as pd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import seaborn as sns
import random
from jupyterthemes import jtplot
jtplot.style(theme="monokai", context="notebook", ticks=True)
```

```python
df = pd.read_csv("../input/fashion-product-images-small/styles.csv", error_bad_lines=False)
df
```

```python
df.articleType.nunique()
```

```python
df.baseColour.nunique()
```

```python
df.season.nunique()
```

```python
df.usage.nunique()
```

```python
df_colors = df.groupby(["baseColour"]).size().reset_index(name="counts").sort_values(by=["counts"], ascending=False)

colors = [f'#{random.randint(0, 0xFFFFFF):06x}' for n in range(df.baseColour.nunique())] # Generate random hex color code

fig, axes = plt.subplots(1, 2, figsize=(15, 10))
sns.barplot(x="baseColour", y="counts", data=df_colors.head(10), palette="hls", ax=axes[0])
axes[1].pie(list(df_colors.counts.values), labels=list(df_colors.baseColour.values), autopct="%1.1f%%", shadow=True, startangle=90, pctdistance=0.85,colors=colors)
center_circle = plt.Circle((0, 0), 0.70, fc='black')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.tight_layout()
plt.show()
```

```python
df_usage = df.groupby(["usage"]).size().reset_index(name="counts").sort_values(by=["counts"], ascending=False)
colors = [f'#{random.randint(0, 0xFFFFFF):06x}' for n in range(df.usage.nunique())] # Generate random hex color code
fig, axes = plt.subplots(1, 2, figsize=(15, 10))
sns.barplot(x="usage", y="counts", data=df_usage,ax=axes[0])
axes[1].pie(list(df_usage.counts.values), labels=list(df_usage.usage.values), autopct="%1.1f%%", shadow=True, startangle=90, pctdistance=0.85,colors=colors)
center_circle = plt.Circle((0, 0), 0.70, fc='black')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.tight_layout()
plt.show()
```

```python
df_season = df.groupby(["season"]).size().reset_index(name="counts").sort_values(by=["counts"], ascending=False)
colors = [f'#{random.randint(0, 0xFFFFFF):06x}' for n in range(df.season.nunique())] # Generate random hex color code
fig, axes = plt.subplots(1, 2, figsize=(15, 10))
sns.barplot(x="season", y="counts", data=df_season, ax=axes[0])
axes[1].pie(list(df_season.counts.values), labels=list(df_season.season.values), autopct="%1.1f%%", shadow=True, startangle=90, pctdistance=0.85,colors=colors)
center_circle = plt.Circle((0, 0), 0.70, fc='black')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.tight_layout()
plt.show()
```
