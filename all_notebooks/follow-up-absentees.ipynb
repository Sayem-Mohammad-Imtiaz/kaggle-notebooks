```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import preprocessing
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('../input/dentalfollowup/Dental_follow-up_attendees.csv')
```

```python
df.head(10)
```

```python
#this dataset is based on the state CO and Zip Code 81137.

#I was checking if the columns had same values or were there any other values.
df['City'].value_counts()
```

```python
#changing categorical values into numerics 

#changing Follow-up Attended?(Yes,No) = (1,0)
df['Follow-up Attended?'].replace(to_replace=['Yes','No'], value=[1,0],inplace=True)
df.head()

#changing City(Allison, Southern Ute) = (1,0)

df['City'].replace(to_replace=['Allison','Southern Ute'], value=[1,0],inplace=True)
df.head()
```

```python
#I'm dropping the colums which have the same data throughout the dataset
df.drop(['State','Zip Code'], axis=1, inplace = True)
```

```python
df.head()
```

```python
df['Marital Status'].value_counts()
```

```python
#I was trying to visualize what number of paitents were attending and not attending the follow-up sessions.
sns.countplot(x='Follow-up Attended?', data = df)
```

```python
#performing an analysis on why paitents are not attending follow ups.

df.rename(columns= {'Follow-up Attended?':'followup'}, inplace = True)
#I changed the column name because it was getting hard for me to remember.
```

```python
#adding a new column to classify the age groups as Youth (15-24 years), Adults (25-64 years)
#Seniors (65 years and over)
l = ['Youth','Adult','Seniors']
newage = []
def new(df,newage):
    for i in range(99):
        if int(df['Age'][i]) in range(15,25):
             newage.append(l[0])
        if int(df['Age'][i]) in range(25,65):
            newage.append(l[1])
        if int(df['Age'][i]) > 65:
            newage.append(l[2])
            
new(df,newage)
```

```python
df = df.assign(Catogary = newage)
```

```python
df.head()
```

```python
#The problem was to figure out why almost 50% of the paitents were not showing up for the followup
#So I seperated the dataset for only the followup absentees 
noatten = df[df['followup'] == 0]

noatten.head()
```

```python
j = noatten['Catogary'].value_counts()
```

```python
#Here we can see that almost 80% of the followup absentees are seniors from ages > 65, so we can see that the 
#problem lies in the age.

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n".format(pct, absolute)




fig, ax = plt.subplots(figsize =(10, 10))
wedges, texts, autotexts = ax.pie(j, 
                                  autopct = lambda pct: func(pct, j), 
                                  labels = j.keys(),
                                  startangle = 90,
                                  colors = ('#A3DDCB','#CAF7E3'),
                                  textprops = dict(color ="black"))
```

```python
tot = df['Catogary'].value_counts()
```

```python
#Here we can also see that more than 50% of the paitents are Seniors. 

def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n".format(pct, absolute)




fig, ax = plt.subplots(figsize =(10, 10))
wedges, texts, autotexts = ax.pie(tot, 
                                  autopct = lambda pct: func(pct, tot), 
                                  labels = tot.keys(),
                                  startangle = 90,
                                  colors = ('#A3DDCB','#CAF7E3','#907FA4'),
                                  textprops = dict(color ="black"))
```
