Создадим функцию которая принимает параметры таблицу Жанр и Язык, возвращает их коллицество. Проверяем колличество фильмов с определённым жанром и языком

```python
def number_film(df, day, city):
    track_list = df[(df['Genre'] == day) & (df['Language'] == city)]
    track_list_count = track_list['IMDB Score'].count()
    return track_list_count
```

```python
number_film(df, 'Documentary', 'English')
```

```python
number_film(df, 'Documentary', 'Hindi')
```

```python
number_film(df, 'Documentary', 'French')
```

```python
number_film(df, 'Documentary', 'Spanish')
```

```python
number_film(df, 'Documentary', 'Italian')
```
