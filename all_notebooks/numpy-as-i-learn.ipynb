## Basic Operations

### Arithmatic Operations

```python
array_17 = np.arange(4)
print(array_17)
array_17 = array_17 + 4  # adding / subtracting scalar value to whole array
print(array_17)
array_17 = array_17 * 3  # multiplying / dividing array by a scalar
print(array_17)
```

Element-wise operations

it is applied only between corresponding elements of operand arrays

```python
array_17 = np.arange(4)
print(array_17)
array_18 = np.arange(8, 12)
print(array_18)
print(array_17 + array_18)
print(array_17 - array_18)
print(array_17 * array_18)
print(array_17 / array_18)
```

```python
array_19 = np.linspace(0, 100, 5)
print(array_19)
array_19 = np.sin(array_19)  # finding sine of each element of the array
print(array_19)
```

```python
array_20 = np.array([2, 4, 8, 16, 32, 64])
print(array_20)
array_20 = np.sqrt(array_20)  # finding square root of each element of the array
print(array_20)
```

Arithmatic operations on multi-dimensional arrays

```python
array_21 = np.arange(0, 9).reshape(3, 3)
print(array_21)
array_22 = np.ones((3, 3))
print(array_22)
print(array_21 + array_22)  # element-wise addition
print(array_21 - array_22)  # element-wise subtraction
print(array_21 * array_22)  # element-wise multiplication
```

### Matrix Product

We have seen until now element-wise multiplication of arrays/matrices. We can also perform matrices multiplication in NumPy as well by using np.dot(), that takes two matrices as arguments.

```python
array_23 = np.arange(0, 9).reshape(3, 3)
print(array_23)
array_24 = np.arange(10, 19).reshape(3, 3)
print(array_24)
array_25 = np.dot(array_23, array_24)
print(array_25)
```

An alternative way of performing same multiplication is writing expression like A.dot(B)

```python
array_26 = array_23.dot(array_24)
print(array_26)
```

Note that since the matrix product is not a commutative operation, the order of the operands is important. Indeed, AxB is not equal to BxA.

```python
array_27 = np.dot(array_24, array_23)
print(array_27)
```

### Increment and Decrement Operators

```python
array_28 = np.arange(5)
print(array_28)
array_28 += 2
print(array_28)
array_28 -= 4
print(array_28)
array_28 *= 3  # multiplying an array with a scalar and saving it in same variable
print(array_28)
```

### Universal Functions (ufunc)

A universal function, generally called ufunc, is a function operating on an array in an element-by-element fashion. This means that it acts individually on each single element of the input array to generate a corresponding result in a new output array. In the end, you obtain an array of the same size as the input.

```python
array_29 = np.arange(3, 8)
print(array_29)
print(np.sqrt(array_29)) # np.sqrt() is ufunc to find square roots of all elements in an array
print(np.log(array_29))  # finds logarithm
print(np.sin(array_29))  # finds sine
# there are many other similar functions, that can be applied on an array element-wise
```

### Aggregate Functions

Aggregate functions perform an operation on an array, and produce a single result. Like sum of all elements of the array.

```python
array_30 = np.arange(13, 20)
print(array_30)
print(array_30.sum())  # finds out sum
print(array_30.min())  # finds out minimum value
print(array_30.max())  # finds out maximum value
print(array_30.mean())  # finds the mean
print(array_30.std())  # finds standard deviation
```

## Indexing, Slicing and Iterating

### Indexing

In order to access a single element of an array, you can refer to its index.

```python
array_31 = np.arange(10, 20)
print(array_31)
print(array_31[1])  # indexing is similar to lists in python
print(array_31[-2])  # negative indexing also applicable
print(array_31[[1, 3, 4, 6, 9]])  # to select multiple indices, we provide list of indices in [<index>]
```

Moving on to two-dimensional arrays, to refer an element we use [row index, col index] as element index

```python
array_32 = np.arange(0, 16).reshape(4, 4)
print(array_32)
print(array_32[1, 1])  # [row index = 1, col index = 1]
print(array_32[-2, -3])  # [row index = -2, col index = -3]
```

### Slicing

```python
array_33 = np.arange(30, 40)
print(array_33)
print(array_33[1:6])  # [<start index> : <end index>]
print(array_33[1:8:2])  # [<start index> : <end index> : <step>]
print(array_33[::2])  # if start and/or end are not provided, it will be 0 and -1 respectively, default step is 1 (i.e. no intervals)
```

For two-dimensional arrays, slicing is applied seperately for rows and for columns

```python
array_34 = np.arange(10, 19).reshape(3, 3)
print(array_34)
print(array_34[0,:])  # printing first row, row = 0, col = all
print(array_34[:,1])  # printing second col, row = all, col = 1
print(array_34[1:3, 1:3])  # printing a part of matrix, with two rows and two cols
print(array_34[[0, 2], 1:3])  # when row/col we want to select are not consecutive, we use an array of wanted indices
```

### Iterating an Array

Iterating is possible using for construct(loop).

```python
array_35 = np.arange(100, 105)
print(array_35)
for i in array_35:  # to print all elements in a single dimensional array
    print(i)
```

```python
array_36 = np.arange(100, 109).reshape(3, 3)
print(array_36)
for i in array_36:  # prints all rows in a double dimensional array
    print(i)
```

```python
array_37 = np.arange(100, 104).reshape(2, 2)
print(array_37)
for i in array_37.flat:  # to make element by element iteration in two-dimensional array
    print(i)
```

### apply_along_axis()

However, despite all this, NumPy offers an alternative and more elegant solution than the for loop.

apply_ along_axis() function takes three arguments: the function, the axis, the array

if we provide a universal function, it will be applied on all the elements of the array seperately

```python
array_38 = np.arange(100, 109).reshape(3, 3)
print(array_38)
print(np.apply_along_axis(np.mean, axis = 0, arr = array_38))  # calculates means of all columns' elements
print(np.apply_along_axis(np.mean, axis = 1, arr = array_38))  # calculates means of all rows' elements
print(np.apply_along_axis(np.sqrt, axis = 1, arr = array_38))  # ufunc, axis does not matter, both return same results
```

we can also create our own fucntions and apply those in apply_along_axis()

```python
array_39 = np.arange(100, 109).reshape(3, 3)
print(array_39)
def half(x):
    return x/2
print(np.apply_along_axis(half, axis = 1, arr = array_39))
```

## Array Manipulation

Often we need to create arrays by using already created arrays

### Joining Arrays

Stacking (vstack() and hstack()) is used to merge two or more arrays to form a new array which will contain all of the elements of all arrays.

```python
array_48 = np.ones((3, 3))
array_49 = np.zeros((3, 3))
array_50 = np.vstack((array_48, array_49))  # vertical stacking of arrays
print(array_50)
array_51 = np.hstack((array_48, array_49))  # horizontal stacking of arrays
print(array_51)
```

Stacking (column_stack() and row_stack()) is used generally with one dimensional arrays which are stacked as columns or rows in order to form a new two dimensional array.

```python
array_52 = np.array([0, 1, 2])
array_53 = np.array([4, 8, 12])
array_54 = np.array([3, 9, 27])
array_55 = np.column_stack((array_52, array_53, array_54))  # cols from one-dim arrays make rows for new array
print(array_55)
array_56 = np.row_stack((array_52, array_53, array_54))  # rows from one-dim arrays make rows for new array
print(array_56)
```

### Splitting Arrays

Splitting (vsplit() and hsplit()) is used to divide arrays.

```python
array_57 = np.arange(0, 16).reshape((4, 4))
print(array_57)
[array_58, array_59] = np.hsplit(array_57, 2)  # array & no. of parts we want array to divide in as arguments
print(array_58)
print(array_59)
```

```python
array_60 = np.arange(0, 16).reshape((4, 4))
print(array_60)
[array_61, array_62] = np.vsplit(array_57, 2)  # splits vertically, meaning height will be divided
print(array_61)
print(array_62)
```

split() allows to divide an array into non-symmetrical parts. It takes three arguments, an array, indexes and axis = 1 or 0 to tell indexes are for cols or for rows

if you want to divide the matrix into three parts, the first of which will include the first column, the second will include the second and the third column, and the third will include the last column, you must specify three indexes in the following way

```python
array_63 = np.arange(0, 16).reshape((4, 4))
print(array_63)
[array_64, array_65, array_66] = np.split(array_63, [1, 3], axis = 1)
print(array_64)
print(array_65)
print(array_66)
```

```python
[array_67, array_68, array_69] = np.split(array_63, [1, 3], axis = 0)
print(array_67)
print(array_68)
print(array_69)
```

## General Concepts

### Copies or Views of Object

When you assign one array a to another array b, you are not copying it; array b is just another way to call array a. Any change in one array will mirror in other array also. The object returned is the view of original array.

```python
array_70 = np.array([1, 2, 3, 4])
array_71 = array_70  # not a copy of an array, just a view
print(array_71)
array_70[2] = 0  # changing an element in array_70
print(array_71)  # change mirrored in array_71
```

Similar is the case when we slice an array and store it in a variable

```python
array_72 = np.arange(5)
array_73 = array_72[1:4]
print(array_73)
array_73[1] = 0
print(array_72)
```

If you want to generate a complete and distinct array, use the copy() function. In this case, if you change one array, other one will remain unchanged.

```python
array_74 = np.array([1, 2, 3, 4, 5])
array_75 = array_74.copy()  # copying array
print(array_75)
array_74[2:4] = [0, 0]  # making a change in one array
print(array_74)  # changed
print(array_75)  # unchanged
```

### Vectorization

Vectorization is the absence of an explicit loop during the developing of the code. These loops actually cannot be omitted, but are implemented internally and then are replaced by other constructs in the code. The application of vectorization leads to a more concise and readable code.

NumPy allows you to express the multiplication of two arrays as shown:

a * b

Or even two matrices:

A * B

In other languages, such operations would be expressed with many nested loops and the for construct.

### Broadcasting

Broadcasting allows an operator or a function to act on two or more arrays to operate even if these arrays do not have the same shape. That said, not all the dimensions can be subjected to broadcasting; they must meet certain rules.

```python
array_76 = np.arange(4)  # single dimensional array
array_77 = np.arange(16).reshape(4, 4)  # two dimensional array
array_76 + array_77  # addition is possible only if there number of cols are same
# an array of (3, 3) shape can not add up with array_76 here
# arrays with (1, 4), (2, 4), (3, 4), etc can
```

Similarly, matrices with different shapes can also be multiplied. But the number of columns of first matrix must be equal to the number of rows of second matrix.

Whenever, we multiply two matrices with (r1, c1) and (r2, c2) shapes, c1=r2 is must and (r1, c2) is the shape of resulted matrix.

```python
array_78 = np.random.random((4, 7))
array_79 = np.random.random((7, 3))
array_80 = np.dot(array_78, array_79)
print(array_80)
```

## Structured Arrays

NumPy allows you to create arrays that are much more complex not only in size, but in the structure, called structured arrays. This type of array contains structs or records instead of individual items.

Each struct can contain bunch of elements with different data types and can be called over by using array indexing.

With the dtype option, you can specify a list of comma-separated specifiers to indicate the elements that will constitute the struct, along with data type and order.

```python
array_81 = np.array([(1, "First", 1.5, 1 + 5j),
                    (2, "Second", 2.5, 2 + 5j),
                    (3, "Third", 3.5, 3 + 5j)],
                   dtype = ("i2, a6, f4, c8"))
array_81
```

You can also use the data type explicitly specifying int8, uint8, float16, complex64, and so forth.

```python
array_82 = np.array([(1, "First", 1.5, 1 + 5j),
                    (2, "Second", 2.5, 2 + 5j),
                    (3, "Third", 3.5, 3 + 5j)],
                   dtype = ("int16, a6, float32, complex64"))
array_82
```

```python
print(array_82[1])  # printing second row which contain a struct of array_82
```

Furthermore, the names assigned to the item of the structs can also be used to call all the elements of similar name in all of the structs. (Like f0, f1, f2 and f3 here)

```python
print(array_82["f1"])
```

As you have just seen, the names are assigned automatically with an f (which stands for field) and a progressive integer that indicates the position in the sequence. In fact, it would be more useful to specify the names with something more meaningful. This is possible and you can do it at the time of array declaration:

```python
array_83 = np.array([(1, "1st", 1.5, 1 + 5j),
                    (2, "2nd", 2.5, 2 + 5j),
                    (3, "3rd", 3.5, 3 + 5j)],
                   dtype = [("integers", "i2"), ("positions", "a6"), ("floats", "f4"), ("complex", "c8")])
array_83
```

```python
print(array_83["positions"])
```

```python
print(array_83.dtype)  # printing dtypes of the structs
print(array_83.dtype.names)  # printing tuple containing names for all dtypes in the structs
```

## Reading and Writing Array Data on File

A very important aspect of NumPy that has not been discussed yet is the process of reading data contained in a file. This procedure is very useful, especially when you have to deal with large amounts of data collected in arrays.

### Loading and Saving Data in Binary Files

functions save() and load() are used to save and later retrieve data stored in binary format.

let's say, we have an array that we have obtained after some work on it and now we want to save it. For this, we use save() function that takes name of the file(string) we're creating and array we want to save as arguments

```python
array_84 = np.array([[0.05512957, 0.73276149, 0.5778144 , 0.70635276],
                     [0.03092344, 0.99958926, 0.63350008, 0.172151  ],
                     [0.21842994, 0.26263767, 0.28870111, 0.29177053],
                     [0.58234664, 0.00869582, 0.69175515, 0.45601734]])
np.save("saved_array", array_84) # created file will automatically have '.npy' extension with it.
```

now, to load the data from saved file or from any other binary file, we will use load() function. We can save the data we are loading into a variable. load() takes file name as argument, if the file is saved in same directory as the program code.

```python
array_85 = np.load("saved_array.npy")
array_85
```

### Reading Files with Tabular Data

Many times, the data that you want to read or save are in textural format (TXT or CSV, for example). You might save the data in this format, instead of binary, because the files can then be accessed outside independently if you are working with NumPy or with any other application.

To be able to read such files, we use genfromtxt(), which takes file name(str), delimiter and whether the data columns are named or not.

```python
array_86 = np.genfromtxt("../input/numpy-as-i-learnt/numpy_AIL.csv", delimiter = ",", names = True)
array_86  # this will give us an structured array, with dtypes and names for all columns
```

One positive aspect of this feature is that if some data are missing, the function can handle them. Let's try to do the same for data that has some removed values.

```python
array_87 = np.genfromtxt("../input/numpy-as-i-learnt/numpy_AIL2.csv", delimiter = ",", names = True)
array_87  # genfromtxt() function replaces the blanks in the file with nan values
```
