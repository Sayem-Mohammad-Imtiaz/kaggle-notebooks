```python
dir1 = '../input/musicnet-dataset/musicnet/musicnet/test_data'
dir2 = '../input/musicnet-dataset/musicnet/musicnet/train_data'
```

```python
from os import listdir
from os.path import isfile, join
files1 = [f for f in listdir(dir1) if isfile(join(dir1, f))]
files2 = [f for f in listdir(dir2) if isfile(join(dir2, f))]
print(len(files1))
print(len(files2))
```

```python
from pydub import AudioSegment
for i in range(0, len(files1)):
    wav_name = dir1 + '/' + files1[i]
    mp3_name = './mp3_outs/' + files1[i] +'.mp3'
    AudioSegment.from_wav(wav_name).export(mp3_name, format="mp3")
    print(i)

for i in range(0, len(files2)):
    wav_name = dir2 + '/' + files2[i]
    mp3_name = './mp3_outs/' + files2[i] +'.mp3'
    AudioSegment.from_wav(wav_name).export(mp3_name, format="mp3")
    print(i)
```

```python
import shutil
shutil.make_archive('mp3_outs', 'zip', './mp3_outs')

import os
file_name = 'mp3_outs.zip'
file_stats = os.stat(file_name)
print(f'File Size in MegaBytes is {file_stats.st_size / (1024 * 1024)}')
```
