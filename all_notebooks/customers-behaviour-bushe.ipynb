# Какие черты Bushe посетителям нравятся по пятибальной шкале?

* Сравнение с другими пекарнями: как бы вы оценили качество Bushe по сравнению с другими брендами?
* Цена. Как бы вы оценили диапазон цен в Bushe?
* Атмосфера: Как бы вы оценили атмосферу в Bushe? (освещение, музыка и т. д.)
* Wi-Fi: вы оцениваете качество Wi-Fi в Bushe как ...
* Сервис: Как бы вы оценили обслуживание в Bushe? (Оперативность, дружелюбие и т. Д.)
* MhRating: Насколько вероятно, что вы выберете Bushe для проведения деловых встреч или встреч с друзьями?

```python
df_new = df_new.rename(columns = 
                       {"12. How would you rate the quality of Bushe compared to other brands (Coffee Bean, Old Town White Coffee..) to be:":"ComparingRate",
                        "13. How would you rate the price range at Bushe?":"PriceRate",
                        "14. How important are sales and promotions in your purchase decision?":"s&pCareness",
                        "15. How would you rate the ambiance at Bushe? (lighting, music, etc...)":"AmbientRate",
                        "16. You rate the WiFi quality at Bushe as..":"WifiRate",
                        "17. How would you rate the service at Bushe? (Promptness, friendliness, etc..)":"ServiceRate",
                        "18. How likely you will choose Bushe for doing business meetings or hangout with friends?":"MhRating",
                        "20. bakery":"bakery",
                        "21. Will you continue buying at Bushe?":"Comeback?"
                        })
df_new.head()
```

```python
# Подсчитаем сумму по каждому столбцу 

Score = {}
Score['Compares']=df_new['ComparingRate'].sum()
Score['Price']=df_new['PriceRate'].sum()
Score['Atmosphere'] = df_new['AmbientRate'].sum()
Score['Wifi'] = df_new['WifiRate'].sum()
Score['Service'] = df_new['ServiceRate'].sum()
Score['MhRating'] = df_new['MhRating'].sum()

# создадим новый   dataframe
score_df = pd.DataFrame(Score.items(), columns=['part','score'])
score_df
```

**Выявлено, что наиболее значимыми положительными факторами приходятся : Атмосферность,Сервис <br>наиболее значимыми не положительными факторами приходятся : Цена**

```python
fig = px.bar(score_df, x='part', y='score')
fig.show()
```

<h3>Оценка персонала, который обслуживал посетителя, пожелавшего зайти снова в пекарню

```python
df1=pd.DataFrame()
df1['comeback'] =df_new['Comeback']
df1['bakery'] = df_new['bakery']
df1['Count'] = (df1[['comeback']] == 'Yes').sum(axis=1)

grouped_df = (
            df1.loc[:, ['bakery','Count']]
            .groupby('bakery')
            .sum().reset_index())

grouped_df
```

```python
fig = px.bar(grouped_df, x='bakery', y='Count')
fig.show()
```

Больше всего посетителей выразили желание вернуться, когда их обслуживал Sebastian
