```python
display (Best2007_2021_df.dtypes)

Best2007_2021_df['DATA'] = pd.to_datetime(Best2007_2021_df['DATA'], format= '%Y%m%d')

display (Best2007_2021_df.dtypes)
```

```python
Best2007_2021_df.plot(x='DATA', y=["AZUL4", "GOLL4", "PETR3", "PRIO3", "USIM3", "VALE3", "CSNA3", "VVAR3", "MGLU3", "ITUB3", "BBDC3"], figsize=(10,10), title=('AÇÕES'), ylabel=('VARIAÇÃO AÇÕES'), xlabel=('YTD') )
```

```python
### MGLU3: Splits:
# 20110502 - 16.45 - MAGALU00 
# 20170904 - 621.79 / 20170905 - 82.0 - Split (desdobramento): 13% / 7,5 Vezes - MAGALU01
# 20190805 - 276.0 / 20190806 - 36.6 - Split (desdobramento): 13% / 7,5 Vezes - MAGALU02
# 20201013 - 104.0 / 20201014 - 25.59 - Split (desdobramento): 25% / 4 Vezes - MAGALU03

# MAGALU00 = MAGALU < 20170905
# MAGALU01 = MAGALU00 * 7,5 : MAGALU < 20190806 & MAGALU > 20170904
# MAGALU02 = MAGALU00 * 7,5 * 7,5 : MAGALU < 20201014 & MAGALU > 20190805
# MAGALU03 = MAGALU00 * 7,5 * 7,5 * 4 : MAGALU > 20201013

```

```python
MAGALU = pd.DataFrame()
#MAGALU00 = MAGALU
#MAGALU01 = MAGALU

MAGALU['MGLU3'] = Best2007_2021_df['MGLU3']
MAGALU['DATA'] = Best2007_2021_df['DATA']

MAGALU
```

```python
# MAGALU00 = MAGALU < 20170905
MAGALU00 = MAGALU.loc[(MAGALU['DATA'] < '2017-09-05')]
MAGALU00
```

```python
# MAGALU01 = MAGALU00 * 7,5 : MAGALU < 20190806 & MAGALU > 20170904
MAGALU01 = MAGALU.loc[(MAGALU['DATA'] < '2019-08-06') & (MAGALU['DATA'] > '2017-09-04')]
MAGALU01['MGLU3'] = MAGALU01['MGLU3'] * 7.5
MAGALU01
```

```python
# MAGALU02 = MAGALU00 * 7,5 * 7,5 : MAGALU < 20201014 & MAGALU > 20190805
MAGALU02 = MAGALU.loc[(MAGALU['DATA'] < '2020-10-14') & (MAGALU['DATA'] > '2019-08-05')]
MAGALU02['MGLU3'] = MAGALU02['MGLU3'] * 56.25
MAGALU02
```

```python
# MAGALU03 = MAGALU00 * 7,5 * 7,5 * 4 : MAGALU > 20201013
MAGALU03 = MAGALU.loc[(MAGALU['DATA'] > '20201013')]
MAGALU03['MGLU3'] = MAGALU03['MGLU3'] * 225
MAGALU03
```

```python
display (MAGALU.shape)
display (MAGALU00.shape)
display (MAGALU01.shape)
display (MAGALU02.shape)
display (MAGALU03.shape)
```

```python
MAGALU00 = MAGALU00.append(MAGALU01)
MAGALU00 = MAGALU00.append(MAGALU02)
MAGALU00 = MAGALU00.append(MAGALU03)
MAGALU00
```

```python
Best2007_2021_df['MGLU3DES'] = MAGALU00 ['MGLU3']
```

```python
svm = Best2007_2021_df.plot(x='DATA', y=["MGLU3DES", "MGLU3"], figsize=(10,10), title=('AÇÕES'), ylabel=('VARIAÇÃO AÇÕES'), xlabel=('YTD') )
figure = svm.get_figure()    
figure.savefig('Magalu.png')
```
