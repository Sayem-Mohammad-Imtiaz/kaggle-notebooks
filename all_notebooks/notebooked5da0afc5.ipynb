## INTRODUCTION

The dataset contains the advertisement budget of companies/organizations spent on mediums such as television (TV), radio, social media & influencers, and the outcome (Sales) for the budget spent. In this analytic, we will see:
* The relationship between the budget & the sale
* Which medium is the best medium for advertising
* The impact of collaboration with influencer

## INSIGHT

From the conclusion of the data analysis, here is some insights can be founded:
* The bigger amount spent on advertisement, the bigger is the sale
* Television is the most effective advertisement medium
* Influencers only effect the advertisement on Social Media medium

```python
#Data Collection

df = pd.read_csv("../input/dummy-advertising-and-sales-data/Dummy Data HSS.csv")

df.info()

df.head()
```

```python
#Checking missing values

missing_values = df.isnull().sum()
missing_values
```

```python
#Dealing with missing values
#Fill the missing values with the mean values

df = df.fillna(df.mean())
df.info()
df.describe()
```

```python
#Converting categorical values to numerical values

df_1 = pd.get_dummies(df, drop_first = True)
df_1.head()
```

```python
sum_column = df["TV"] + df["Radio"] + df["Social Media"]
df["Total Budget"] = sum_column
print(df)
```

## Explaratory Data Analysis

```python
#Data Visualization

df_1.hist()
```

```python
plt.title = ('Total Budget Spent vs Sales')
plt.figure(figsize=(14,7))
sns.barplot(x=df["Total Budget"], y=df["Sales"])
```

The bar chart above shows that the bigger the budget spent, the bigger is the outcome, although it doesn't happen in some cases. This is possibly due to:
* The advertisement/campaign effectiveness
* The choice of medium

```python
plt.title = ("Budget for TV vs Sales")
plt.figure(figsize=(14,7))
sns.barplot(x=df['TV'], y=df["Sales"])
```

The bar chart above shows the strong correlation between the budget for TV and sales. This shows that TV is still one of the best medium for advertising.

```python
plt.figure(figsize=(14,7))
sns.scatterplot(x=df['TV'], y=df['Sales'], hue=df['Influencer'])
```

The scatter plot above show the influence of Influencer somehow doesn't gives a big impact to the advertisement on TV.

```python
plt.title = ("Budget for Radio vs Sales")
plt.figure(figsize=(14,7))
sns.barplot(x=df['Radio'], y=df["Sales"])
```

The bar chart 'Budget for Radio vs Sales' above shows almost similar pattern with the 'Budget for TV vs Sales', but somehow in some cases, even a small or medium size budget can produce a bigger outcome. 

```python
plt.figure(figsize=(14,7))
sns.scatterplot(x=df['Radio'], y=df['Sales'], hue=df['Influencer'])
```

The scatter plot above shows the influence of Influencer somewhat give some impact for the advertisement for radio.

```python
plt.title = ("Budget for Social Media vs Sales")
plt.figure(figsize=(14,7))
sns.barplot(x=df['Social Media'], y=df["Sales"])
```

The bar chart 'Budget for Social Media vs Sales' shows that the amount of budget spent doesn't affect too much on the outcome.

```python
plt.figure(figsize=(14,7))
sns.scatterplot(x=df['Social Media'], y=df['Sales'], hue=df['Influencer'])
```

The scatter plot above shows that the Influencer give a big impact on the outcome. Althought a small budget spent on social media, but with the influence of Influencer, the outcome can be very high.

```python
plt.figure(figsize=(14,7))
sns.swarmplot(x=df['Influencer'], y=df['Sales'])
```

The swarm plot above shows that whatever budget spent on the influencer (scale from Nano to Mega), doesn't promise a big outcome.

## CONCLUSION

From the data analized above, here we can conclude that:
* The bigger amount spent on advertisement, the bigger is the sale
* Television is the most effective advertisement medium
* Influencers only effect the advertisement on Social Media medium
