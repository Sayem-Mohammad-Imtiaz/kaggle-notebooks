```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('../input/flow-of-financial-resources/RBI data.csv')
data.head()
```

```python
data.rename(columns={'Y-o-Y Growth in Adjusted Credit in percentage':'Y-o-Y Growth'},inplace=True)
data.rename(columns={'Non-food Bank Credit':'Credit'},inplace=True)
data.rename(columns={'Investments in Commercial Paper':'Inv in Comm. Paper'},inplace=True)
data.rename(columns={'Investments in Shares':'Inv in shares'},inplace=True)
data.rename(columns={'Investments in Bonds/Debentures':'Inv in bonds'},inplace=True)
data.rename(columns={'Total non SLR Investments':'Total invs'},inplace=True)
```

```python
data.rename(columns={'Adjusted Non-food Bank Credit':'Adjusted credit'},inplace=True)
```

```python
data.head()
```

```python
data.set_index('Fortnight Ended')
```

```python
fig,axes=plt.subplots(1,2,figsize=(20,5))
sns.histplot(data['Credit'],color='pink',ax=axes[0])
sns.histplot(data['Adjusted credit'],color='grey',ax=axes[1])
```

```python
plt.figure(figsize=(15,4))
plt.scatter(x=data['Credit'],y=data['Adjusted credit'],color='orange',alpha=0.5)
```

```python
plt.figure(figsize=(30,10))
plt.xticks(rotation=90)
sns.lineplot(x=data['Fortnight Ended'],y=data['Y-o-Y Growth'],color='red')
```
