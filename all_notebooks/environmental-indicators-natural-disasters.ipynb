---
Environmental indicators are simple measures that tell us what is happening in the environment. Since the environment is very complex, indicators provide a more practical and economical way to track the state of the environment than if we attempted to record every possible variable in the environment.
    
A few environment indicators in the dataset are: -
*   Air and Climate
*   Biodiversity
*   Forests
*   Natural Disaster and many more.
    
In this notebook we’ll focus only about a single indicator i.e. Natural Disaster. A natural disaster is a major adverse event resulting from natural processes of the Earth; examples include floods, hurricanes, tornadoes, volcanic eruptions, earthquakes, tsunamis, storms, and other geologic processes. A natural disaster can cause loss of life or damage property, and typically leaves some economic damage in its wake, the severity of which depends on the affected population's resilience and on the infrastructure available. 
    
Natural Disasters can be classified into: -
* Climatological
* Geophysical
* Hydrological
* Meteorological

```python
import numpy as np 
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
import matplotlib.gridspec as gridspec
import seaborn as sns

sns.set(font="Serif")
sns.set_style('darkgrid')
sns.set(rc={'axes.facecolor':'whitesmoke','grid.color':'#003e1f','figure.facecolor':'white'})


params = {'legend.fontsize': 'x-large',
          'figure.figsize': (20, 32),
         'axes.labelsize': 'large',
         'axes.titlesize':'large',
         'xtick.labelsize':'large',
         'ytick.labelsize':'large'}
pylab.rcParams.update(params)
import warnings
warnings.filterwarnings("ignore")
```

<a style="font-family: Verdana; font-size: 16px; font-style: normal; font-weight: bold; letter-spacing: 3px; color: #003E1F; background-color: #ffffff;">Ploting Function</a>

```python
def top10(df,col):
    df = df.dropna(subset=[col])
    df[col]=df[col].astype(str).astype(int)
    df=df.sort_values(by=col,ascending=False)[:10]
    df=df.sort_values(by=col,ascending=True)
    return df

def plot_3_columns(df,col1,col2,col3,color):
    fig, ax = plt.subplots(2,2,figsize=(30, 16))
    df1 = top10(df,col1)
    df2 = top10(df,col2)
    df3 = top10(df,col3)
    
    gs = gridspec.GridSpec(2, 7)
    gs.update(wspace = 0.1, hspace = 0.3)
    
    ax0 = plt.subplot(gs[0, :3])
    sns.stripplot(x=col1,y="Countries or areas",data=df1,linewidth=5,marker="*",size=15,edgecolor=color,alpha=.50,ax=ax0)
    
    ax1 = plt.subplot(gs[0, 4:])
    ax1.set_xscale('log')
    sns.stripplot(x=col2,y="Countries or areas",data=df2,marker="o",size=15,edgecolor=color,color=color,alpha=.50,ax=ax1)
    
    ax2 = plt.subplot(gs[1, 1:5])
    ax2.set_xscale('log')
    sns.stripplot(x=col3,y="Countries or areas",data=df3,marker="D",size=20,edgecolor=color,color=color,alpha=.50,ax=ax2)

    plt.show()
```

<a style="font-family: Verdana; font-size: 24px; font-style: normal; font-weight: bold; letter-spacing: 3px; color: #003E1F; background-color: #ffffff;">EDA for Climatological Diaster</a>

<a style="font-family: Verdana; font-size: 16px; font-style: normal; font-weight: bold; letter-spacing: 3px; color: #696773; background-color: #ffffff;">Importing data</a>

```python
climatological_df = pd.read_csv("../input/global-environmental-indicators/Natural Disasters/Climatological disasters.csv")
climatological_df.replace('...', np.nan, inplace=True)
climatological_df.head()
```

<a style="font-family: Verdana; font-size: 16px; font-style: normal; font-weight: bold; letter-spacing: 3px; color: #696773; background-color: #ffffff;">Year 1990-1999</a>

```python
plot_3_columns(climatological_df,"Occurrence 1990-1999","Total deaths 1990-1999","Persons affected 1990-1999","#696773")
```

From the above graph we can see that China Special Region in 1990-1999 had most no. of Climatological natural disaster. But India despite having comparatively less no. of natural disaster more people were affected in India and since more people are affected by the calamity it only makes sense that India had most no. of deaths in 1990-1999.

From the year 2000-2009 we can see that United states of America had more than 30 climatological natural disaster still they were able to cope up well to save their people lives as they are not in top 10 count of people affected nor in the count of people deaths.

```python
plot_3_columns(climatological_df,"Occurrence 2010-2019","Total deaths 2010-2019","Persons affected 2010-2019","#696773")
```

From all the above graphs we can observe that more no. citizens of India and China were affected even though not having that many climatological disasters in the span of years. As China and India leads the population graphs of the world even a few natural disasters compared to other countries can affect more people. As these countries are densely populated.
