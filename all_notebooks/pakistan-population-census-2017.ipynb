```python
df['District'] = df['District'].str.replace('FR ', '')
df['District'] = df['District'].str.replace(' DISTRICT', '')
df['District'] = df['District'].str.strip()
df['District'] = df['District'].str.lower()
df['Tehsil'] = df['Tehsil'].str.replace(' TEHSIL', '')
df['Tehsil'] = df['Tehsil'].str.replace(' SUB-TEHSIL', '')
df['Tehsil'] = df['Tehsil'].str.replace(' CITY', '')
df['Tehsil'] = df['Tehsil'].str.lower()
df['Tehsil'] = df['Tehsil'].str.strip()
district = df['District'].tolist()
tehsil = df['Tehsil'].tolist()
District_Tehsil = list(set(district+tehsil))
df.sample(50)
```

```python
district = df['District'].tolist()
tehsil = df['Tehsil'].tolist()
District_Tehsil = list(set(district+tehsil))
District_Tehsil
```
