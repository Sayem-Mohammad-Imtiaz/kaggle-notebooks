# Emphasizing continuity with line plots


<div class="alert alert-block alert-info">üìç With some datasets, you may want to understand changes in one variable as a function of time, or a similarly continuous variable. In this situation, a good choice is to draw a line plot. In seaborn, this can be accomplished by the lineplot() function. There is one more method to do this.</div>

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script', cursive;"> ‚≠êÔ∏è  kind="line" </span>

```python
sns.relplot(x="year", y="price", kind="line", data=data).set_xticklabels(rotation=30);
```

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script',cursive;">Because lineplot() assumes that you are most often trying to draw y as a function of x, the default behavior is to sort the data by the x values before plotting. However, this can be disabled:</span>

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script', cursive;"> ‚≠êÔ∏è  sort=False </span>

```python
sns.relplot(x="year", y="price",sort=False ,kind="line", data=data).set_xticklabels(rotation=30);
```

# Aggregation and representing uncertainty

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script',cursive;"> More complex datasets will have multiple measurements for the same value of the x variable. The default behavior in seaborn is to aggregate the multiple measurements at each x value by plotting the mean and the 95% confidence interval around the mean:

```python
sns.relplot(x="year", y="mileage", kind="line", data=data).set_xticklabels(rotation=30);
```

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script',cursive;">The confidence intervals are computed using bootstrapping, which can be time taking for larger datasets. So we can disable them setting ci=None

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script', cursive;"> ‚≠êÔ∏è  ci=None </span>

```python
sns.relplot(x="year", y="mileage",ci=None,kind="line", data=data).set_xticklabels(rotation=30);
```

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script',cursive;">Another good option, especially with larger data, is to represent the spread of the distribution at each timepoint by plotting the standard deviation instead of a confidence interval:

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script', cursive;"> ‚≠êÔ∏è  ci="sd" </span>

```python
sns.relplot(x="year", y="mileage",ci="sd",kind="line", data=data).set_xticklabels(rotation=30);
```

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script',cursive;">To turn off aggregation altogether, set the estimator parameter to None This might produce a strange effect when the data have multiple observations at each point.

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script', cursive;"> ‚≠êÔ∏è  estimator=None </span>

```python
sns.relplot(x="year", y="mileage",estimator=None,kind="line", data=data).set_xticklabels(rotation=30);
```

# Plotting subsets of data with semantic mappings


<div class="alert alert-block alert-info">üìå The lineplot() function has the same flexibility as scatterplot(): it can show up to three additional variables by modifying the hue, size, and style of the plot elements</div>

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script',cursive;"> Using semantics in lineplot() will also determine how the data get aggregated. For example, adding a hue semantic with two levels splits the plot into two lines and error bands, coloring each to indicate which subset of the data they correspond to:

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script', cursive;"> ‚≠êÔ∏è  Using hue  
 ‚≠êÔ∏è  kind="line" </span>

```python
sns.relplot(x="year", y="mileage",hue="drive",kind="line", data=data).set_xticklabels(rotation=30);
```

```python
sns.relplot(x="year", y="mileage",hue="drive",kind="line", data=data).set_xticklabels(rotation=30);
```

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script',cursive;"> Adding a style semantic to a line plot changes the pattern of dashes in the line by default:

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script', cursive;"> ‚≠êÔ∏è  Using Style </span>

```python
sns.relplot(x="year", y="mileage",hue="drive",style="body",kind="line", data=data).set_xticklabels(rotation=30);
```

But you can identify subsets by the markers used at each observation, either together with the dashes or instead of them:

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script', cursive;"> ‚≠êÔ∏è  setting markers=True </span>

```python
sns.relplot(x="year", y="mileage",hue="drive",style="body",markers=True,kind="line", data=data).set_xticklabels(rotation=30);
```

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script',cursive;"> But you can identify subsets by the markers used at each observation, either together with the dashes or instead of them:

```python
sns.relplot(x="year", y="mileage",hue="drive",style="body",dashes=False,markers=True,kind="line", data=data).set_xticklabels(rotation=30);
```

# Showing multiple relationships with facets


<span style="font-size:18px;color:crimson;font-family: 'Dancing Script',cursive;"> The best approach may be to make more than one plot. Because relplot() is based on the FacetGrid, this is easy to do. To show the influence of an additional variable, instead of assigning it to one of the semantic roles in the plot, use it to ‚Äúfacet‚Äù the visualization. This means that you make multiple axes and plot subsets of the data on each of them: </span>

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script', cursive;"> ‚≠êÔ∏è  Using col="feature_name" </span>

```python
sns.relplot(x="year", y="price", hue="drive",
            col="body", data=data).set_xticklabels(rotation=30);
```

<span style="font-size:16px;color:gray;">You can also show the influence two variables this way: one by faceting on the columns and one by faceting on the rows. As you start adding more variables to the grid, you may want to decrease the figure size. Remember that the size FacetGrid is parameterized by the height and aspect ratio of each facet:

<span style="font-size:18px;color:crimson;font-family: 'Dancing Script', cursive;"> ‚≠êÔ∏è  Using col="feature_name"   
‚≠êÔ∏è  Using row="feature_name"   
‚≠êÔ∏è  Setting Height </span>

```python
sns.relplot(x="year",y="mileage",hue="body",col="drive",row="engType",height=3,kind="line", estimator=None,data=data).set_xticklabels(rotation=30);
```
