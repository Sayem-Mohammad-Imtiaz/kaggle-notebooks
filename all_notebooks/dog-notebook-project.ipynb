```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = {}
data['2007'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2007.csv')
data['2008'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2008.csv')
data['2009'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2009.csv')
data['2010'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2010.csv')
data['2011'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2011.csv')
data['2012'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2012.csv')
data['2013'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2013.csv')
data['2014'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2014.csv')
data['2015'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2015.csv')
data['2016'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2016.csv')
data['2017'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2017.csv')
```

# the most common breeds in 2017

```python
(data['2017']).Breed.value_counts()
```

# looking for a beagle named gunner in 2017

```python
data['2017'].loc[(data['2017'].Breed == 'BEAGLE') & (data['2017'].DogName == 'GUNNER')]
```

# spefifying what i want in the filtered data (beagles and what colors they should be)

```python
data_filtered=data['2017'].loc[(data['2017'].Breed=='BEAGLE') & (data['2017'].Color.isin(
    ['WHITE/BLACK/BROWN','WHITE/BROWN','BROWN'])),["Breed","LicenseType","DogName","Color"]]
data_filtered 
```

# the most common colors among the specified beagles

```python
plt.plot(data_filtered.Color.value_counts())
```

# the most common license types for beagles of a specific color

```python
data_filtered.LicenseType.value_counts()
```

# the most common names for beagles of specific colors

```python
fig=plt.figure()
ax1=fig.add_subplot()
ax1.set_xlabel('DogName')
ax1.set_ylabel('Count')
ax1.set_title('Distribution of names for Beagles')
ax1.bar(data_filtered.DogName.value_counts()[:8].index.array,data_filtered.DogName.value_counts()[:8])
```

# the 5 most common colors among beagles and labs 

```python
lab_beagle = data['2015'].loc[(data['2015'].Breed =='BEAGLE') | (data['2015'].Breed == 'LAB MIX')]
lab_beagle.Color.value_counts()[:5]
```

# the 15 most common names of labs, goldens and chihuahuas'

```python
DOG = data['2007'].loc[(data['2007'].Breed == 'LAB MIX') | (data['2007'].Breed == 'GOLDEN RETRIEVER') | (data['2007'].Breed == 'CHIHUAHUA')]
DOG.DogName.value_counts()[:15]
```

# finding out the most popular breeds in 2013 and 2011 combined...

```python
doge = (data['2013'].Breed.value_counts()[:10]) + (data['2011'].Breed.value_counts()[:10])
doge = doge.sort_values(ascending=False)
doge

fig=plt.figure(figsize=(20,5))
ax1=fig.add_subplot()
ax1.set_xlabel('Breed')
ax1.set_ylabel('Count')
ax1.set_title('numbers of certain Breeds in 2007 and 2013')
ax1.bar(doge.index.array,doge.values)
```
