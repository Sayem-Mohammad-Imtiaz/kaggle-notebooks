```python
data = pd.read_csv("/kaggle/input/turkish-employee-dataset/calisan_data.csv")

data.head()
```

```python
#conversion turkish month names to english
#then datetime object
for i in range(len(data)):
    date = data["DogumTarihi"][i]
    date = date.replace("Ocak","January")
    date = date.replace("Şubat","February")
    date = date.replace("Mart","March")
    date = date.replace("Nisan","April")
    date = date.replace("Mayıs","May")
    date = date.replace("Haziran","June")
    date = date.replace("Temmuz","July")
    date = date.replace("Ağustos","August")
    date = date.replace("Eylül","September")
    date = date.replace("Ekim","October")
    date = date.replace("Kasım" ,"November")
    date = date.replace("Aralık","December")
    
    date = datetime.datetime.strptime(date,"%d %B %Y")
    
    data["DogumTarihi"][i] = date
    
#conversion datetime to float
#data["DogumTarihi"] = dts.date2num(data_clean["DogumTarihi"])

data.head()
```

```python
plt.figure(figsize=(7,7))
plt.title("Çalışanların Departmanlara Dağılımı")
plt.axis("equal")
data.groupby(['Departman']).size().plot(kind='pie', autopct='%1.1f%%')
plt.show()
plt.figure(figsize=(7,7))
plt.title("Çalışanların Şehirlere Dağılımı")
plt.axis("equal")
data.groupby(['Sehir']).size().plot(kind='pie', autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x = "Departman",y = "Maas",data = data)
plt.show()
```

```python
department_color = []

for i in data["Departman"]:
    if i == "Yazılım Geliştirme":
        department_color.append("Blue")
    elif i == "Mali İşler":
        department_color.append("Orange")
    elif i == "Müşteri İlişkileri ve Satış":
        department_color.append("Green")
    elif i == "İnsan Kaynakları":
        department_color.append("Red")
    elif i == "Yönetim":
        department_color.append("Purple")

p = [mpatches.Patch(color='Blue', label='Yazılım Geliştirme'),mpatches.Patch(color='Orange', label='Mali İşler'),
     mpatches.Patch(color='Green', label='Müşteri İlişkileri ve Satış'),mpatches.Patch(color='Red', label='İnsan Kaynakları'),
     mpatches.Patch(color='Purple', label='Yönetim')]

plt.figure(figsize = (10,10))
plt.scatter(dts.date2num(data["DogumTarihi"])/365,data["Maas"],c = department_color,label = departments)
plt.xlabel("Doğum Tarihi")
plt.ylabel("Maaş")
plt.legend(handles = p)
plt.show()
```
