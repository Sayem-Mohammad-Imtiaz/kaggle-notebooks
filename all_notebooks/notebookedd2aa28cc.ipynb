```python
train_data = pd.read_csv("/kaggle/input/house-price-prediction-challenge/train.csv")
```

```python
test_data = pd.read_csv("/kaggle/input/house-price-prediction-challenge/test.csv")
```

```python
print(train_data.info())
```

```python
X_train = train_data.iloc[:, :-1].values
```

```python
y_train = train_data.iloc[:, -1].values
```

```python
X_test = X_test = test_data.iloc[:, :].values
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X_train[:, 0] = le.fit_transform(X_train[:, 0])

le2 = LabelEncoder()
X_train[:, 4] = le2.fit_transform(X_train[:, 4])

le3 = LabelEncoder()
X_train[:, 8] = le3.fit_transform(X_train[:, 8])

le4 = LabelEncoder()
X_test[:, 0] = le4.fit_transform(X_test[:, 0])

le5 = LabelEncoder()
X_test[:, 4] = le5.fit_transform(X_test[:, 4])

le6 = LabelEncoder()
X_test[:, 8] = le6.fit_transform(X_test[:, 8])
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=50, random_state=0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
file = open("my_prediction.csv", "w")
file.write("TARGET(PRICE_IN_LACS)\n")
for i in regressor.predict(X_test):
    file.write(str(i)+"\n")   
file.close()
```
