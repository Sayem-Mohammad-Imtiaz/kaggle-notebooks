# Applying Model

```python
label=df["Result"]
data=df.drop('Result',axis=1)

from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(data,label)
```

# Model Score

```python
model.score(data, label)
```

# Accuracy Score

```python
from sklearn.metrics import accuracy_score
accuracy_score(label, model.predict(data)) 
```

# Mean Square Error

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(label, model.predict(data)) 
```

# The value of MSE is near to 0 and not more. So, It have less errors. 

```python
df1 = pd.DataFrame({'Actual': label, 'Predicted': model.predict(data), 'Difference': round( (label - model.predict(data)),2 ) }) 
df1.head(10)
```
