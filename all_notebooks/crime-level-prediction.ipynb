```python
df = data
df = df.fillna(0)
```

```python
df.drop("CHAS", axis = 1,inplace = True)
scaler = StandardScaler()
X = scaler.fit_transform(df.drop(['CRIM'], axis = 1))
y = scaler.fit_transform(df[['CRIM']])

```

```python
model = LinearRegression()
lin_reg_model_pred = cross_val_predict(model, X, y, cv=5)
scores_linear = cross_val_score(model, X, y, cv=5, scoring="neg_mean_squared_error")
mse_linear = mean_squared_error(y, lin_reg_model_pred)
print (f"linear cross_val_scores: {scores_linear}")
print (f"linear mse_score:        {mse_linear:.10}")
```
