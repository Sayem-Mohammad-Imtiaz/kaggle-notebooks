> IMPORTING DATASET

```python
dset=pd.read_csv('/kaggle/input/gufhtugu-publications-dataset-challenge/GP Orders - 4.csv');dset
```

#  EXPLORING DATA (CHECKING NULL VALUES)

```python
#column name rewrite for avoid erorrs
columns = dset.columns.str.replace(' ','_').str.lower()
columns = columns.str.replace('[(,)]', '')
dset.columns = columns
print(dset.columns)
```

# Reterieved total no of rows and colums


```python
dset.head()
print(dset.shape)
dset.info()
```

# Checking Null Values & Fixing

```python
#check if any entity in cloumns is null
print(dset.isnull().sum())
#if any entry is  null replace it by most occuring entry
dset.book_name=dset.book_name.fillna(dset.book_name.mode()[0])
dset.city_billing=dset.city_billing.fillna(dset.city_billing.mode()[0])
print("\n*-----------------after removing null values------------*\n")
print(dset.isnull().sum())
```

```python
#print Book Names seprated by "/"
dset.book_name = dset.book_name.str.split('/')
print(dset.book_name)

dset['order_date'] = pd.to_datetime(dset['order_date'])
print(dset.head())
```

# Data Exploratory Analysis

**1. ORDER STATUS FREQUENCY**

```python
#HISTOGRAM
ord_sts=dset.order_status.unique()
print(ord_sts)
print("\n *-----------Histogram of ORDER STATUS-----------------*")
plt.style.use('ggplot')
plt.title('Order STATUS')
plt.xlabel('STATUS')
plt.ylabel('No Of Orders')
plt.hist(dset.order_status)
plt.show()
```

Total List of years

```python
# Retrieve Total List of years 
years = dset['order_date'].dt.strftime("%Y").unique().tolist()
print(years)
```

ORDER STATUS PER YEAR  

```python
# reterive order status in 2019
sts_19 = dset[dset['order_date'].dt.strftime("%Y") == "2019"].order_status
# reterive order status in 2020
sts_20 = dset[dset['order_date'].dt.strftime("%Y") == "2020"].order_status
# reterive order status in 2021
sts_21 = dset[dset['order_date'].dt.strftime("%Y") == "2021"].order_status
# graph size
plt.figure(figsize=(20,8))
#histogram for 2019
plt.subplot(1,3,1)
plt.title('Order Status in 2019')
plt.xlabel('Order status')
plt.ylabel('Total no of orders')
plt.hist(sts_19)

#histogram for 2020
plt.subplot(1,3,2)
plt.title('Order Status in 2020')
plt.xlabel('Order status')
plt.ylabel('Total no of orders')
plt.hist(sts_20)

#histogram for 2021
plt.subplot(1,3,3)
plt.title('Order Status in 2021')
plt.xlabel('Order status')
plt.ylabel('Total no of orders')
plt.hist(sts_21)
```

**2. Best Selling Books**

```python
#use split method(/) for seprating books
book_data = dset.book_name.apply(lambda x: str(x).split('/'))
# Storing the list of lists book names in a list 'books' 
books = [item for sublist in book_data for item in sublist]
# making an new data set for ease
dset1 = pd.DataFrame(data = books, columns = ['NoOfSoldBooks'])
# Only storing the top 10 most selling books of all times and converting the result into a datafram
book_chart = dset1.NoOfSoldBooks.value_counts().nlargest(15).to_frame()
# Printing the most sold book name and its number of times it was sold
print(book_chart.head(1))
print()
# plotting 10 top selling books of all time
px.bar(book_chart, y = book_chart.NoOfSoldBooks, x = book_chart.index, title = 'Best Selling Books')
```

#  City Billing in MOST CITES

```python
#books sells in top most 8 cities
most_cities = dset['city_billing'].value_counts()
most_cities = most_cities.iloc[:8]
#pi-chart of top  most cities where books sold
most_cities.plot(kind='pie',figsize=(8,8),autopct='%1.1f%%')
```
