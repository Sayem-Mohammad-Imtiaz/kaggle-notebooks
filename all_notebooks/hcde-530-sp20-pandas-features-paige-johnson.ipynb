Now create a new variale with a pandas data series, including the null value (NaN).

Show what the variable s contains:

Pandas contains a date_range() feature that allows you to generate any arbitrary range of dates for you to work with. Let's create a variable `dates` that contains a range of dates in 6 date periods and then display its contents.

```python
dates = pd.date_range('20200101', periods = 6)
dates
```

We can add column labels to these dates and put some numbers into the cells of the dataframe. Since we don't have any numbers handy, we will use the numpy random number generator to fill in those cells. Then we display the table with all of the values filled in.

```python
df = pd.DataFrame(np.random.randn(6,4), index=dates, columns=list('ABCD'))
df
```

Remember you can show the last rows of a table with the `tail()` function.

```python
df.tail(1)
```

The `index()` function shows you what the column labels are. This is useful if you want to select out ranges or columns of data with other pandas functions.

```python
df.index
```

You can use the `describe` function to display some simple statistics. This is not so useful for randomly generated numbers, but with real data it an be quite handy.

```python
df.describe()
```

Suppose you want to swap the rows and columns. This is what we'd call a pivot table in Excel. You can reflect the DataFrame over its main diagonal by writing rows as columns and vice-versa. The property `T` is an *accessor* to the method `transpose()`. Note that the results are that the columns are now rows and vice-versa.

```python
df.T
```

BUt note also that the dataframe itself is unchanged. (If you wanted to store the transposed results above, you'd need to store it as a variable.)

```python
df
```

Recall that the integer locate method `iloc` allows us to select out particular rows and columns. You could select [1,1] to do a further analysis.

```python
df.iloc[[1,1,1,1,1,1,2,3,4]]
```

## Simple plot
You can plot a simple visualizaiton using building plotting fetaures in the `plot()` function (which is part of the matplotlib library). The next code generates a time series `ts` of random values using the techniques described above. It then plots a cumulative summary of the values in the series using `plot()`. Plot has tons of styling features that can be modified. Check the documentation for a list of those: https://matplotlib.org/tutorials/introductory/pyplot.html

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2020', periods=1000))
cum = ts.cumsum()
cum.plot()
```
