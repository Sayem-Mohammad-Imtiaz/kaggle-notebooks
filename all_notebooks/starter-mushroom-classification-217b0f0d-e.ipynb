```python
from sklearn import preprocessing
```

```python
enc = preprocessing.OrdinalEncoder()
enc.fit(df)
```

```python
tf = enc.transform(df)
```

```python
type(tf)
```

```python
x = tf[:,1:]
```

```python
y = tf[:,0]
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
c = LinearRegression()
```

```python
c.fit(x,y)
```

```python
c.score(x,y)
```

```python
from sklearn import svm
c2 = svm.SVC()
c2.fit(x,y)
c2.score(x,y)
```
