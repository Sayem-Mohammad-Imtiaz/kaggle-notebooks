```python
#delete rows where country is nan
wines2 = wines.dropna(subset=['country'])
```

```python
#check
len(wines2.country.unique())
```

```python
#show overall point distribution
wines2.points.plot.hist(bins=12, alpha=0.5)
```

**Well, this distribution is almost normal. But there are a lack of 89pts wines. Maybe our experts intend to give some wines 90 pts instead of 89 in order to increase wines' category?
Now I will not trust such wines and prefer to choose bottle with 88-89, as more honest.**

```python
#show point distribution by country
for country in wines2.country.unique():
    wines2[wines2.country == country].points.plot.hist(bins=12, alpha=0.5, figsize=(18,5))
```

```python
#the same on separate subplots
fig, axes = plt.subplots(11, 4, figsize=(15,25), dpi=100, sharex=True, sharey=True)
for i, (ax, country) in enumerate(zip(axes.flatten(), wines2.country.unique())):
    x = wines2.loc[wines2.country==country, 'points']
    ax.hist(x, alpha=0.5, bins=12, density=True, stacked=True, label=str(country))
    ax.set_title(country)
```

```python
#delete countries with less than 50 values, then repeat
wines3 = wines2[wines2.groupby('country')['country'].transform('count').gt(50)]
fig, axes = plt.subplots(7, 4, figsize=(15,25), dpi=100, sharex=True, sharey=True)
for i, (ax, country) in enumerate(zip(axes.flatten(), wines3.country.unique())):
    x = wines3.loc[wines3.country==country, 'points']
    ax.hist(x, alpha=0.5, bins=12, density=True, stacked=True, label=str(country))
    ax.set_title(country)
```

```python
#plot the rest, exclude countries with only 1 wine
wines4 = wines2[wines2.groupby('country')['country'].transform('count').le(50)]
wines4 = wines4[wines4.groupby('country')['country'].transform('count').gt(1)]
fig, axes = plt.subplots(4, 4, figsize=(15,15), dpi=100, sharex=True, sharey=True)
for i, (ax, country) in enumerate(zip(axes.flatten(), wines4.country.unique())):
    x = wines4.loc[wines4.country==country, 'points']
    ax.hist(x, alpha=0.5, bins=12, density=True, stacked=True, label=str(country))
    ax.set_title(country)
```
