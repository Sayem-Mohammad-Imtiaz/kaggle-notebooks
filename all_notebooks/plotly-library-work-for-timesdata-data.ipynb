# Line Chart
Citation and Teaching vs Word Rank of Top 100 Universities year 2011

```python
df = timesData.iloc[:100,:]
import plotly.graph_objs as go         #Kütüphaneyi import eder

    #Plotly kütüphanesinden scatter metodunu çağır.

trace1 = go.Scatter(
                    x = df.world_rank,         #x ekseni, dünya sıralaması (1-100arasında)
                    y = df.citations,          #y ekseni, citation skoru (1-100 arasında değişiyor)
                    mode = "lines",            #plotun şekli line olarak seçildi
                    name = "citations",        #ilk yaratacağımız trace citations
                    marker = dict(color ='rgba(240,112,2,0.8)'),  # renk, -0,8saydamlık oranı   
                    text = df.university_name)  #plotun üzerine gelince görmek istediğim yazı,burda üniversite adı yazılıyor.
    
    
    
trace2 = go.Scatter(
                    x = df.world_rank,         #x ekseni, dünya sıralaması (1-100arasında)
                    y = df.teaching,           #y ekseni, teaching skoru   (1-100 arasında değişiyor)
                    mode = "lines+markers",    #plotun şekli line+marker olarak seçildi
                    name = "teaching",
                    marker = dict(color ='rgba(110,26,80,0.8)'),
                    text = df.university_name)
data = [trace1, trace2]                        #data listesi içerisinde her ikisini birleştir

# x ve y ekseni üzerinde işlemler yapalım
layout = dict(title = 'Citation and Teaching vs Word Rank of Top 100 Universities',
              xaxis = dict(title = 'World Rank',ticklen= 2,zeroline= False)   
             )
#layout ve datayı fig içerisinde birleştiriyoruz
fig = dict(data = data, layout = layout)
iplot(fig)   #figürü çizdirir
```

# Scatter Plot
Scatter Example: Citation vs world rank of top 100 universities with 2014, 2015 and 2016 years

```python
#timeData mı al timeData'nin year'ını 2014'e eşitle ve bunun ilk 100 üniversitesini al ve df oluştur.

df2014 = timesData[timesData.year == 2014].iloc[:100,:]
df2015 = timesData[timesData.year == 2015].iloc[:100,:]
df2016 = timesData[timesData.year == 2016].iloc[:100,:]
#import graph object as "go"

import plotly.graph_objs as go

trace1 = go.Scatter(
                    x = df2014.world_rank,
                    y = df2014.citations,
                    mode = "markers",
                    name = "2014",
                    marker = dict(color = 'rgba(255, 120, 200, 0.8)'),
                    text= df2014.university_name)
trace2 = go.Scatter(
                    x = df2015.world_rank,
                    y = df2015.citations,
                    mode = "markers",
                    name = "2015",
                    marker = dict(color = 'rgba(254, 126, 3, 0.8)'),
                    text= df2015.university_name)
trace3 = go.Scatter(
                    x = df2016.world_rank,
                    y = df2016.citations,
                    mode = "markers",
                    name = "2016",
                    marker = dict(color = 'rgba(0, 128, 216, 0.8)'),
                    text= df2016.university_name)

data = [trace1, trace2, trace3]
layout = dict(title = 'Citation vs world rank of top 100 universities with 2014, 2015 and 2016 years',
              xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'Citation',ticklen= 5,zeroline= False)
             )
fig = dict(data = data, layout = layout)
iplot(fig)
```

# Bar Charts
2014 yılındaki ilk 3 üniversitenin citations ve teachinglerini karşılaştırma 

```python
#dataframe filtre hazırlama
df2014 =timesData[timesData.year == 2014].iloc[:3,:]
#plotly.graph kütüphanesini go olarak import edelim
import plotly.graph_objs as go

trace1 = go.Bar(
                    x = df2014.university_name,
                    y = df2014.citations,
                    name = "citations",
                    marker = dict(color = 'rgba(255, 174, 255, 0.8)',
                                 line=dict(color='rgb(0,0,0)',width=1.5)),
                    text = df2014.country)
trace2 = go.Bar(
                    x = df2014.university_name,
                    y = df2014.teaching,
                    name = "teaching",
                    marker = dict(color = 'rgba(255, 80, 90, 0.8)',
                                 line=dict(color='rgb(0,0,0)',width=1.5)),
                    text = df2014.country)
data = [trace1, trace2]
layout= go.Layout(barmode = "group") #blokları yan yana grup olarak yapar
fig= go.Figure(data = data, layout = layout)
iplot(fig)
```

# Bubble Chart
Renge ve size anlamlar katar
İlk 20 üniversitenin teaching skorunu üniversite rengine göre karşılaştırılacak.Number of student(size),international score(color) olacak.

```python
df2016 = timesData[timesData.year == 2016].iloc[:20,:]
num_students_size  = [float(each.replace(',', '.')) for each in df2016.num_students]
international_color = [float(each) for each in df2016.international]
data = [
    {
        'y': df2016.teaching,
        'x': df2016.world_rank,
        'mode': 'markers',
        'marker': {
            'color': international_color,
            'size': num_students_size,
            'showscale': True
        },
        "text" :  df2016.university_name    
    }
]
iplot(data)
```

# Histogram
2011-2012 yıllarındaki öğrenci-personel oranı

```python
x2011 = timesData.student_staff_ratio[timesData.year == 2011]
x2012 = timesData.student_staff_ratio[timesData.year == 2012]

trace1 = go.Histogram(
    x = x2011,
    opacity=0.75,
    name = "2011",
    marker=dict(color='rgba(171,50,96,0.6)'))
trace2= go.Histogram(
    x=x2012,
    opacity=0.75,
    name = "2012",
    marker=dict(color='rgba(12,50,196,0.6)'))
data = [trace1, trace2]
layout = go.Layout(barmode = 'overlay',
                   title ='student-staff ratio in 2011 and 2012',
                   xaxis =dict(title='student-staff-ratio'),
                   yaxis =dict(title='Count'),
)
fig = go.Figure(data = data, layout = layout)
iplot(fig)
                   
```

# Scatter Plot
2 tane feature arasındaki ilişkiyi görmek için kullanılır.

```python
# import figure factory
import plotly.figure_factory as ff
# prepare data
dataframe = timesData[timesData.year == 2015]
data2015 = dataframe.loc[:,["research","international", "total_score"]]
data2015["index"] = np.arange(1,len(data2015)+1)
# scatter matrix
fig = ff.create_scatterplotmatrix(data2015, diag='box', index='index',colormap='Portland',
                                  colormap_type='cat',
                                  height=700, width=700)
iplot(fig)
```

# Inset Plots
Üniversitelerin teaching score ve income karşılaştır.


```python
# first line plot
trace1 = go.Scatter(
    x=dataframe.world_rank,
    y=dataframe.teaching,
    name = "teaching",
    marker = dict(color = 'rgba(255, 255, 2, 0.8)'),
)
# second line plot
trace2 = go.Scatter(
    x=dataframe.world_rank,
    y=dataframe.income,
    xaxis='x2',
    yaxis='y2',
    name = "income",
    marker = dict(color = 'rgba(50, 112, 20, 0.8)'),
)
data = [trace1, trace2]
layout = go.Layout(
    xaxis2=dict(
        domain=[0.6, 0.95],
        anchor='y2',        
    ),
    yaxis2=dict(
        domain=[0.6, 0.95],
        anchor='x2',
    ),
    title = 'Income and Teaching vs World Rank of Universities'

)

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

# 3D Scatter Plot with Colorscaling


```python
# create trace 1 that is 3d scatter
trace1 = go.Scatter3d(
    x=dataframe.world_rank,
    y=dataframe.research,
    z=dataframe.citations,
    mode='markers',
    marker=dict(
        size=10,
        color='rgb(155,22,2)',   # set color to an array/list of desired values     
       
    )
)

data = [trace1]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0  
    )
    
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
