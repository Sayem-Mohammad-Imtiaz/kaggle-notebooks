# Create a global animation of COVID vaccinations (Total number of people fully vaccinated per hundred)
Total number of people fully vaccinated per hundred is chosen to visuallize, in order to see the percentage of fully vaccinated process of each country in global scale.

```python
import numpy as np 
import pandas as pd
import plotly.express as px

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
print(df.columns)
fig = px.choropleth(df,locations='country',locationmode='country names',color='people_fully_vaccinated_per_hundred',hover_name='country',animation_frame='date',color_continuous_scale='reds',range_color=[0,50])
fig.update_layout(title_text='Global animation of fully COVID vaccinations by percentage',title_x=0.4)
fig.show()
```

# Create a global map of earthquake occurences worldwide
From 2008-05-10 00:00:00 to 2008-05-16 23:59:59, in which the 2008 Sichuan Earthquake occured on May 12, 2008
https://www.theatlantic.com/photo/2018/05/10-years-since-the-devastating-2008-sichuan-earthquake/560066/

```python
df1 = pd.read_csv('../input/2008-sichuan-earthquake-3days-before-and-after/query (1).csv',header=0)
print(df1.columns)
df1.index = pd.to_datetime(df1['time'])
df1['time'] = df1.index.strftime('%Y-%m-%d %H:00:00')
df1 = df1.sort_index()
fix1 = px.scatter_geo(df1,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='turbo',range_color=[1,8])
fix1.show()
```
