```python
import pandas as pd
import datetime
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
# Function to calculate daily deltas from aggregated data
def get_deltas(orig_series):
    result_series = [0]
    i = 0
    while i < orig_series.shape[0] - 1:
        delta = orig_series[i+1] - orig_series[i]
        if delta >= 0:
            result_series.append(delta)
        else:
            result_series.append(0)
        i+=1
    return result_series
```

```python
data = pd.read_csv("../input/belarus-covid19-stats/belarus_covid19_data.csv")

for column in data.drop("date_pub", axis=1).columns:
    data[column + "_dayly"] = get_deltas(data[column])
```

```python
fig, ax = plt.subplots(2,2, figsize=(40,10));

step = 22
plt.suptitle("COVID-19 in Belarus Dayly Stats", fontsize=16)

ax[0][0].bar(data["date_pub"], data["tested_dayly"]);
ax[0][0].set_ylim(0);
ax[0][0].set_title("Tested");
ax[0][0].set_xticks(data["date_pub"][::step])

ax[0][1].bar(data["date_pub"], data["infected_dayly"]);
ax[0][1].set_ylim(0);
ax[0][1].set_title("Infected");
ax[0][1].set_xticks(data["date_pub"][::step]);

ax[1][0].bar(data["date_pub"], data["recovered_dayly"]);
ax[1][0].set_ylim(0);
ax[1][0].set_title("Recovered");
ax[1][0].set_xticks(data["date_pub"][::step]);

ax[1][1].bar(data["date_pub"], data["died_dayly"]);
ax[1][1].set_ylim(0);
ax[1][1].set_title("Died");
ax[1][1].set_xticks(data["date_pub"][::step]);

plt.tight_layout()
```
