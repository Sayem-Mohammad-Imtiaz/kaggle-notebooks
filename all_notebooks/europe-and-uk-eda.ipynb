```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 

plt.style.use('fivethirtyeight')
plt.rcParams["figure.figsize"] = (16,5)
```

```python
data_1979_2021 = pd.read_csv("../input/monthly-gold-prices/1979-2021.csv")
data_1990_2021 = pd.read_csv("../input/monthly-gold-prices/1990-2021.csv")
```

# EDA of monthly prices from 1979 to 2021 

```python
data_1979_2021.head()
```

```python
data_1979_2021.info()
```

```python
data_1979_2021.describe()
```

## Missing Values

```python
cols_with_missing = [col for col in data_1979_2021.columns if data_1979_2021[col].isnull().any()]
print("Number of columns with missing values : " , len(cols_with_missing))
```

```python
cols_with_missing = [col for col in data_1979_2021.columns if data_1979_2021[col].isnull().any()]
sum_of_missing = data_1979_2021.isnull().sum()[cols_with_missing]

plt.bar(cols_with_missing ,sum_of_missing )
plt.xlabel("Countries")
plt.ylabel("Count of Missing Values");
```

## PLOTS
