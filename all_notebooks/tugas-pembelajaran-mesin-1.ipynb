```python
#Pertama, buat variabel x,y
x = df.drop(columns= 'Y_harga_unit')
y = df['Y_harga_unit']
```

```python
#Kedua, Kita split data kita menjadi training and testing dengan porsi 80 : 20
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
#Ketiga, kita bikin object linear regresi
lin_reg = LinearRegression()
```

```python
#Kempat, uji coba model menggunakanj training data yang sudh di split
lin_reg.fit(x_train, y_train)
```

```python
#Kelima, cari tau nilai slope (m) dan intercerpt (b)
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
coef_dict = {
    'features' : x.columns,
    'coef_value' : lin_reg.coef_
}
coef = pd.DataFrame(coef_dict, columns=['features', 'coef_value'])
coef
```

```python
y_prend = lin_reg.predict(x_test)
```

```python
#Kenam, kita cari tahu akurasi skor dari model yang kita gunakan testing data yang sudah di slpit
lin_reg.score(x_test, y_test)
```

```python
# Pediksi harga real estate 
# X1_Usia_Rumah X2_jarak_MRT X3_jumlah_toko
lin_reg.predict([[2,100,2]])
```
