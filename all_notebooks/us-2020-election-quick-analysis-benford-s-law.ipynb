```python
Benford_check = pd.DataFrame()

Benford_check['Biden'] = president_counties.loc[president_counties.candidate == 'Joe Biden'].total_votes.astype(str).str[0].astype(int).value_counts()
Benford_check['Trump'] = president_counties.loc[president_counties.candidate == 'Donald Trump'].total_votes.astype(str).str[0].astype(int).value_counts()
Benford_check['Benford'] = [30.1, 17.6, 12.5, 9.7, 7.9, 6.7, 5.8, 5.1, 4.6, 0.0]

Benford_check.Biden = Benford_check.Biden / Benford_check.Biden.sum() * 100
Benford_check.Trump = Benford_check.Trump / Benford_check.Trump.sum() * 100
```

```python
Benford_check.plot.bar(color={'Biden': 'blue','Trump': 'red', 'Benford': 'purple'})
#Benford_check.Trump.plot(kind='bar',legend=True,color='red')
#Benford_check.Benford.plot()

```
