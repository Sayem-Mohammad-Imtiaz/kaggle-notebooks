```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
```

*lalu kita panggil, dataset yang akan kita gunakan....*

```python
data_train = pd.read_csv('/kaggle/input/main-tenis/play_tennis_train.csv')
data_test = pd.read_csv('/kaggle/input/main-tenis/play_tennis_test.csv')
```

*selanjutnya kita akan split datset..*

```python
from sklearn.model_selection import train_test_split
```

```python
le = preprocessing.LabelEncoder()
data_train_df = pd.DataFrame(data_train)
data_train_df_encoded = data_train_df.apply(le.fit_transform)
data_test_df = pd.DataFrame(data_test)
data_test_df_encoded = data_test_df.apply(le.fit_transform)
```

*berikut data test...*

```python
x_test = data_test_df_encoded.drop(['play'],axis=1)
y_test = data_test_df_encoded['play']
print(x_test)
```

*dan dibawah adalah data train...*

```python
x_train = data_train_df_encoded.drop(['play'],axis=1)
y_train = data_train_df_encoded['play']
print(x_train)
```

*lalu kita masuk ke proses KNN...*

```python
knn_ = np.arange(1,9)
train_accuracy =np.empty(len(knn_))
test_accuracy = np.empty(len(knn_))

for i,k in enumerate(knn_):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    train_accuracy[i] = knn.score(x_train, y_train)
    test_accuracy[i] = knn.score(x_test, y_test) 
```

```python
plt.title('KNN in Descision Tree')
plt.plot(knn_, test_accuracy, label='Test')
plt.plot(knn_, train_accuracy, label='Train')
plt.legend()
plt.xlabel('Number of neighbors')
plt.ylabel('Accuracy')
plt.show()
```
