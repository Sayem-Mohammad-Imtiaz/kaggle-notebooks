```python
# Import essential libraries
from matplotlib import pyplot as plt
import seaborn as sns
from matplotlib import cm
import pylab as pl
from pandas.plotting import scatter_matrix, andrews_curves, parallel_coordinates, lag_plot, autocorrelation_plot, bootstrap_plot, radviz, table
```

```python
# Reading the dataset
data_2015 = pd.read_csv('../input/world-happiness/2015.csv')
data_2016 = pd.read_csv('../input/world-happiness/2016.csv')
data_2017 = pd.read_csv('../input/world-happiness/2017.csv')
data_2018 = pd.read_csv('../input/world-happiness/2018.csv')
data_2019 = pd.read_csv('../input/world-happiness/2019.csv')
```

```python
# Understanding one of the dataset
# data_2015.head(2)
data_2015
# data_2015.info()
# data_2015[['Happiness Rank', 'Happiness Score', 'Standard Error', 'Economy (GDP per Capita)','Family','Health (Life Expectancy)','Freedom','Trust (Government Corruption)','Generosity','Dystopia Residual']]
```

```python
# Histogram depiction for the Generosity over years
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0.5, hspace=1)
figure, axes = plt.subplots(1,5,figsize=(20,5))
plt.subplot(151); plt.hist(x=data_2015.Generosity, bins='auto', color='#0504aa',alpha=0.7, rwidth=0.85); plt.title('Generosity, 2015')
plt.subplot(152); plt.hist(x=data_2016.Generosity, bins='auto', color='#aaaa00',alpha=0.7, rwidth=0.85); plt.title('Generosity, 2016')
plt.subplot(153); plt.hist(x=data_2017.Generosity, bins='auto', color='#00e504',alpha=0.7, rwidth=0.85); plt.title('Generosity, 2017')
plt.subplot(154); plt.hist(x=data_2018.Generosity, bins='auto', color='#a0504a',alpha=0.7, rwidth=0.85); plt.title('Generosity, 2018')
plt.subplot(155); plt.hist(x=data_2018.Generosity, bins='auto', color='#efc1ff',alpha=0.7, rwidth=0.85); plt.title('Generosity, 2019')
pl.suptitle("Generosity Distribution over years (2015-2019)")
```

```python
# Freedom for making life choices
plt.subplots_adjust(left=0, bottom=0, right=10, top=10, wspace=1, hspace=1)
figure, axes = plt.subplots(1,5, figsize=(15,4))
data_2015.plot.box(y='Freedom', color = '#bb2299', ax=axes[0]); plt.title('Freedom in 2015')
data_2016.plot.box(y='Freedom', color = '#229933', ax=axes[1]); plt.title('Freedom in 2016')
data_2017.plot.box(y='Freedom', color = '#0337bd', ax=axes[2]); plt.title('Freedom in 2017')
data_2018.plot.box(y='Freedom to make life choices', color = '#882200', ax=axes[3]); plt.title('Freedom in 2018')
data_2019.plot.box(y='Freedom to make life choices', color = '#d049ff', ax=axes[4]); plt.title('Freedom in 2019')
```

```python
# Kernel Density Estimate plot
figure, axes = plt.subplots(5, 3, figsize=(15,15))
data_2015.plot.kde(y='Happiness Score', bw_method=0.3, color='#aac4dd', ax=axes[0,0])
data_2015.plot.kde(y='Happiness Score', bw_method=1, color='#c4aadd', ax=axes[0,1])
data_2015.plot.kde(y='Happiness Score', bw_method=3, color='#ddaac4', ax=axes[0,2])

data_2016.plot.kde(y='Happiness Score', bw_method=0.3, color='#aac4dd', ax=axes[1,0])
data_2016.plot.kde(y='Happiness Score', bw_method=1, color='#c4aadd', ax=axes[1,1])
data_2016.plot.kde(y='Happiness Score', bw_method=3, color='#ddaac4', ax=axes[1,2])

data_2017.plot.kde(y='Happiness.Score', bw_method=0.3, color='#aac4dd', ax=axes[2,0])
data_2017.plot.kde(y='Happiness.Score', bw_method=1, color='#c4aadd', ax=axes[2,1])
data_2017.plot.kde(y='Happiness.Score', bw_method=3, color='#ddaac4', ax=axes[2,2])

data_2018.plot.kde(y='Score', bw_method=0.3, color='#aac4dd', ax=axes[3,0])
data_2018.plot.kde(y='Score', bw_method=1, color='#c4aadd', ax=axes[3,1])
data_2018.plot.kde(y='Score', bw_method=3, color='#ddaac4', ax=axes[3,2])

data_2019.plot.kde(y='Score', bw_method=0.3, color='#aac4dd', ax=axes[4,0])
data_2019.plot.kde(y='Score', bw_method=1, color='#c4aadd', ax=axes[4,1])
data_2019.plot.kde(y='Score', bw_method=3, color='#ddaac4', ax=axes[4,2])
```

```python
# Hexagonal binning plot
figure, axes = plt.subplots(1, 5, figsize=(20,5))
data_2015.plot.hexbin(x='Dystopia Residual', y= 'Family', C = 'Happiness Score', reduce_C_function=np.sum, gridsize=20, cmap= "viridis", ax=axes[0])
data_2016.plot.hexbin(x='Dystopia Residual', y= 'Family', C = 'Happiness Score', reduce_C_function=np.sum, gridsize=20, cmap= "YlOrRd", ax=axes[1])
data_2017.plot.hexbin(x='Dystopia.Residual', y= 'Family', C = 'Happiness.Score', reduce_C_function=np.sum, gridsize=20, cmap= "ocean", ax=axes[2])
data_2018.plot.hexbin(x='Perceptions of corruption', y= 'Social support', C = 'Score', reduce_C_function=np.sum, gridsize=20, cmap= "rainbow", ax=axes[3])
data_2019.plot.hexbin(x='Perceptions of corruption', y= 'Social support', C = 'Score', reduce_C_function=np.sum, gridsize=20, cmap= "terrain", ax=axes[4])
```

```python
# Scatter plot
plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0.5, hspace=1)
figure, axes = plt.subplots(1, 5, figsize=(20,4))
data_2015.plot.scatter(x='Family',y='Freedom',c='Magenta', s = 'Happiness Score', marker=5, cmap="viridis", alpha=0.3, linewidths=3, edgecolors='#ff00ff', ax=axes[0])
data_2016.plot.scatter(x='Family',y='Freedom',c='Green', s = 'Happiness Score', marker=8, cmap="viridis", alpha=0.4, linewidths=3, edgecolors='#ffff00', ax=axes[1])
data_2017.plot.scatter(x='Family',y='Freedom',c='Red', s = 'Happiness.Score', marker=3, cmap="viridis", alpha=0.5, linewidths=3, edgecolors='#00ffff', ax=axes[2])
data_2018.plot.scatter(x='Social support',y='Freedom to make life choices',c='Purple', s = 'Score', marker=4, cmap="viridis", alpha=0.6, linewidths=3, edgecolors='#ff0000', ax=axes[3])
data_2019.plot.scatter(x='Social support',y='Freedom to make life choices',c='Blue', s = 'Score', marker=6, cmap="viridis", alpha=0.7, linewidths=3, edgecolors='#00ff00', ax=axes[4])
```
