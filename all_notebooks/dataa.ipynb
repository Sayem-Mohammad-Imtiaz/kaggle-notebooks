**MATPLOTLİB**

   Matplot is a python library that help us to plot data. The easiest and basic plots are line, scatter and histogram plots.
* Line plot is better when x axis is time.
* Scatter is better when there is correlation between two variables
* Histogram is better when we need to see distribution of numerical data.
* Customization: Colors,labels,thickness of line, title, opacity, grid, figsize, ticks of axis and linestyle

```python
#Line plot
data.symmetry_worst.plot(kind = 'line', color = 'black',label = 'symmetry_worst',linewidth=1,alpha = 0.7,grid = True,linestyle = ':')
data.smoothness_mean.plot(kind = 'line', color = 'orange', label = 'smoothness_mean', linewidth=1, alpha = 0.7, grid = True)
data.fractal_dimension_se.plot(kind = 'line', color = 'blue', label = 'fractal_dimension_se', linewidth=1, alpha=0.7, grid = True, linestyle = '-.')

plt.legend()
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("Line_Plot")
plt.show()
```

```python
#Scatter Plot
plt.scatter(data.symmetry_worst, data.smoothness_mean, color="red",label ="data")
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.show()
```

DICTIONARY
Why we need dictionary?

It has 'key' and 'value'
Faster than lists 
What is key and value. Example:
dictionary = {'spain' : 'madrid'}
Key is spain.
Values is madrid. 

It's that easy. 
Lets practice some other properties like keys(), values(), update, add, check, remove key, remove all entries and remove dicrionary.


```python
dictionary = {'black' : 'siyah','green' : 'yeşil','pink' : 'pembe', 'red' : 'kırmızı','blue' : 'mavi'}
print(dictionary.keys())
print(dictionary.values())
```

```python
#update existing entry
dictionary['black'] = 'En koyu renk'
print(dictionary)
```

```python
#add new entry
dictionary['purple'] = 'mor'
print(dictionary)
```

```python
# remove entry
del dictionary['green']
print(dictionary)
```

```python
print('purple' in dictionary)
```

```python
#remove all entries dictionary
dictionary.clear()
print(dictionary)


```
