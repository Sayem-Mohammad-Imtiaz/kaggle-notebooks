# Importing Libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing dataset

```python
dataset=pd.read_csv("../input/kc-house-data/kc_house_data.csv")
dataset.head()
```

# Data Preprocessing 

```python
dataset.describe() #gives you entire stats about all the indexes.
```

```python
#Treating Outliers
bed=dataset['bedrooms']
pr=dataset['price']
plt.title("No. Bedroom vs Price")
plt.xlabel("No of Bedrooms")
plt.ylabel("Price")
plt.scatter(bed,pr)
 #no of bedroom=33 is an outlier which can be removed
```

```python
# check for the extreme values shown in the describe.
dataset=dataset[dataset['bedrooms']<33]

#Splitting into x and y dataset
y=dataset['price']
x= dataset.drop(columns=["price","id","lat","long"])

```

```python
xdf=x

# Deriving new columns
xdf['year_old']=2020-xdf['yr_built']
xdf['renovated_status'] = np.where(xdf['yr_renovated']>0, 1, 0)
xdf=xdf.drop(columns=['date','sqft_living15','sqft_lot15','yr_renovated','yr_built'])

xdf.isnull().sum() #2 records in sqft_above is null
```

```python
#treating null values 
xdf=xdf.replace(np.nan,0)
```

# Splitting dataset


```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(xdf,y,test_size=0.2)
```

# Building Model

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
lin_reg1=LinearRegression()
lin_reg1.fit(xtrain,ytrain)
```

```python
#Polynomial Regression
#converting x to polynomial to check the status of polynomial regression
from sklearn.preprocessing import PolynomialFeatures
polyreg=PolynomialFeatures(degree=2)
xpoly=polyreg.fit_transform(xtrain)
lin_reg2=LinearRegression()
lin_reg2.fit(xpoly,ytrain)
```

# Testing & Anallyzing the Results
#MultiLinear
#Polynomial

```python
ypred1 = lin_reg1.predict(xtest) #output from Linear Regression
ypred2=  lin_reg2.predict(polyreg.fit_transform(xtest)) #Output from Polynomial Regression
from sklearn.metrics import r2_score
score1= r2_score(ytest,ypred1)#Score from Linear Regression
score2=r2_score(ytest,ypred2)#Score from Polynomial Regression
print('Score for MultiLinear Regression',str(score1),'\n','Score for Polynomial Regression',str(score2))

```

```python
#boosting accuracy
from xgboost import XGBRegressor
regboost=XGBRegressor(learning_rate=0.5, max_depth=4, n_estimators=500)
regboost.fit(xtrain,ytrain)
```

```python
yboost=regboost.predict(xtest)
from sklearn.metrics import r2_score
score3= r2_score(ytest,yboost)
print("Score with XGBoost ",score3)
```
