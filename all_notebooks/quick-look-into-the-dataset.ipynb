```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("../input/cannabis-data/supply.csv")
data["Indicator"].unique()
```

We can try to use the supplies data to see how it has varied over time!
firstly we will try to seperate out the points which are produced by Domestic markets and then we will also seperate our data into those that are for medical use and those that are for non medical use.

```python
y = data[(data["Indicator"]=="Domestic market production")& (data["Type of use"]=="Medical use")]["VALUE"]
```

```python
x = data[(data["Indicator"]=="Domestic market production")& (data["Type of use"]=="Medical use")]["REF_DATE"]
```

```python
plt.plot(x,y)
```

Hmmm there is a clear rise in supplies for medicinal use after 2010

```python
y = data[(data["Indicator"]=="Domestic market production")& (data["Type of use"]=="Non-medical use")]["VALUE"]
```

```python
x = data[(data["Indicator"]=="Domestic market production")& (data["Type of use"]=="Non-medical use")]["REF_DATE"]
```

```python
plt.plot(x,y)
```

```python
y = data[(data["Indicator"]=="Domestic market production")& (data["Type of use"]=="Medical use")]["VALUE"]
```

hmm although there is a few drops there , we can clearly observe the increase in supplies of cannabis for non medicinal purposes,  how about we check illegal supplies from abroad too:

```python
y = data[(data["Indicator"]=="Illegal purchases from abroad")& (data["Type of use"]=="Non-medical use")]["VALUE"]
x = data[(data["Indicator"]=="Illegal purchases from abroad")& (data["Type of use"]=="Non-medical use")]["REF_DATE"]
```
