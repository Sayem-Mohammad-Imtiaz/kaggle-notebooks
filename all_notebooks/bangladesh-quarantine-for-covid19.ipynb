```python
nRowsRead = 1000 # specify 'None' if want to read whole file
df = pd.read_csv('../input/cusersmarildownloadsbquarantinecsv/bquarantine.csv', delimiter=';', encoding = "ISO-8859-1", nrows = nRowsRead)
df.dataframeName = 'bquarantine.csv'
nRow, nCol = df.shape
print(f'There are {nRow} rows and {nCol} columns')
df.head()
```

```python
#Code from Nitin Datta https://www.kaggle.com/nitindatta/eda-with-r3-id/comments

fig,ax=plt.subplots(figsize=(18,6))
sns.set_style('dark')
plt.subplot(1, 2, 1)
g1=sns.countplot(df['Division'],palette='rocket', hue = df['Geometry'],**{'hatch':'/','linewidth':3})
g1.set_title("Division")

plt.subplot(1, 2, 2)
g2=sns.countplot(df['Geometry'],palette='rocket',hue= df['Geometry'],**{'hatch':'/','linewidth':3})
g2.set_title("Geometry")
```

```python
#Code by Olga Belitskaya https://www.kaggle.com/olgabelitskaya/parts-of-speech

plt.figure(figsize=(10,5))
sns.countplot(y="Division",data=df,
             facecolor=(0,0,0,0),linewidth=5,
             edgecolor=sns.color_palette("RdYlGn"))
plt.title('Division Quarantine for Covid-19',
         fontsize=15);
```

```python
#Code by Taha07  https://www.kaggle.com/taha07/data-scientists-jobs-analysis-visualization/notebook

color = plt.cm.RdGy(np.linspace(0,1,20))
df["Division"].value_counts().sort_values(ascending=False).head(20).plot.pie(y="total_quarantine",colors=color,autopct="%0.1f%%")
plt.title("")
plt.axis("off")
plt.show()
```

```python
import plotly.offline as pyo
import plotly.graph_objs as go
lowerdf = df.groupby('Distname').size()/df['Shape_Area'].count()*100
labels = lowerdf.index
values = lowerdf.values

# Use `hole` to create a donut-like pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values,marker_colors = px.colors.sequential.speed, hole=.6)])
fig.show()
```

```python
fig = px.pie(df,
             values="total_quarantine",
             names="Division",
             template="seaborn")
fig.update_traces(rotation=90, pull=0.05, textinfo="percent+label")
fig.show()
```

```python
# Count Plot
plt.style.use("classic")
plt.figure(figsize=(10, 8))
sns.countplot(df['Division'], palette='RdYlGn')
plt.xlabel("Division")
plt.ylabel("Count")
plt.title("Bangladesh Quarantine for Covid-19")
plt.xticks(rotation=45, fontsize=8)
plt.show()
```

#To change the Marker Color we write `marker_color` or `marker_line_color`. Just color results in error.

```python
fig = go.Figure();
fig.add_trace(go.Scatter(x = df['Division'].head(10),y = df['total_quarantine'],
                    mode='lines+markers',
                    name='total_quarantine'));

fig.update_traces(mode='lines+markers', marker_line_width=2, marker_size=10, marker_color='red');

fig.update_layout(autosize=False, width=1000,height=700,title="Bangla Divisions Quarantine for Covid-19", legend_orientation="h");

fig.show();
```

```python
fig = go.Figure();
fig.add_trace(go.Scatter(x = df['Distname'].head(10),y = df['total_quarantine'],
                    mode='lines+markers',
                    name='total_quarantine'));

fig.update_traces(mode='lines+markers', marker_line_width=2, marker_size=10, marker_color='green');

fig.update_layout(autosize=False, width=1000,height=700,title="Bangladesh Quarantine for Covid-19", legend_orientation="h");


fig.show();
```

```python
fig = px.treemap(df, path=['Division'], values='total_quarantine',
                  color='total_quarantine', hover_data=['Division'],title="Bangladesh Divisions Quarantine for Covid-19",
                  color_continuous_scale='RdYlGn')
fig.show()
```

```python
#word cloud
from wordcloud import WordCloud, ImageColorGenerator
text = " ".join(str(each) for each in df.Division)
# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=200,colormap='Reds', background_color="green").generate(text)
plt.figure(figsize=(10,6))
plt.figure(figsize=(15,10))
# Display the generated image:
plt.imshow(wordcloud, interpolation='Bilinear')
plt.axis("off")
plt.figure(1,figsize=(12, 12))
plt.show()
```
