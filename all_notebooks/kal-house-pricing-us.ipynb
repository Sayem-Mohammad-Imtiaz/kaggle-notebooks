```python

fig, axes = plt.subplots(2,2, figsize=(12,10)) 
data_train.GarageCond.value_counts().sort_index().plot.bar(ax=axes[0][0]) 
axes[0][0].set_title("GarageCond", fontsize=18)

data_train.GarageQual.value_counts().sort_index().plot.bar(ax=axes[0][1]) 
axes[0][1].set_title("GarageQual", fontsize=18)

data_train.GarageFinish.value_counts().sort_index().plot.bar(ax=axes[1][0]) 
axes[1][0].set_title("GarageFinish", fontsize=18)

data_train.GarageType.value_counts().sort_index().plot.bar(ax=axes[1][1]) 
axes[1][1].set_title("GarageType", fontsize=18)
```

```python
data_train['GarageCond'] = data_train['GarageCond'].fillna('None') 
data_train['GarageQual'] = data_train['GarageQual'].fillna('None') 
data_train['GarageType'] = data_train['GarageType'].fillna('None') 
data_train['GarageFinish'] = data_train['GarageFinish'].fillna('None') 
```

```python
fig, axis = plt.subplots(2,2, figsize = (12,10))
sns.boxplot(data=data_train, x = 'GarageCond', y = 'SalePrice', ax=axis[0,0])
axis[0,0].set_title('GarageCond Vs Sale Price' )

sns.boxplot(data=data_train, x = 'GarageQual', y = 'SalePrice', ax=axis[0,1])
axis[0,1].set_title('GarageQual Vs Sale Price' )

sns.boxplot(data=data_train, x = 'GarageType', y = 'SalePrice', ax=axis[1,0])
axis[1,0].set_title('GarageType Vs Sale Price' )

sns.boxplot(data=data_train, x = 'GarageFinish', y = 'SalePrice', ax=axis[1,1])
axis[1,1].set_title('GarageFinish Vs Sale Price')

plt.tight_layout()
```

So, what did we learn from the above plot?

The house sale price increases; if the quality of basment is excellent/good.
The house sale price increases; if the basment condition is good.

```python
fig, axis = plt.subplots(2,2, figsize = (12,10))

sns.violinplot(data=data_train, x = 'Electrical', y = 'SalePrice', ax=axis[0,0])
axis[0,0].set_title('Electrical Vs Sale Price' )

sns.boxplot(data=data_train, x = 'Heating', y = 'SalePrice', ax=axis[0,1])
axis[0,1].set_title('Heating Vs Sale Price' )

sns.boxplot(data=data_train, x = 'HeatingQC', y = 'SalePrice', ax=axis[1,0])
axis[1,0].set_title('HeatingQC Vs Sale Price' )

sns.violinplot(data=data_train, x = 'CentralAir', y = 'SalePrice', ax=axis[1,1])
axis[1,1].set_title('CentralAir Vs Sale Price')

plt.tight_layout()
```

# So, we can see from the above plot that having Excellent Heating Quality and Central Air Conditioning increase the sale prices.

```python
fig, axis = plt.subplots(2,2, figsize = (14,12))

sns.boxplot(data=data_train, x = 'SaleCondition', y = 'SalePrice', ax=axis[0,0])
axis[0,0].set_title('SaleCondition Vs Sale Price' )

sns.boxplot(data=data_train, x = 'Condition2', y = 'SalePrice', ax=axis[0,1])
axis[0,1].set_title('Condition2  Vs Sale Price' )

sns.boxplot(data=data_train, x = 'Condition1', y = 'SalePrice', ax=axis[1,0])
axis[1,0].set_title('Condition1 Vs Sale Price' )

sns.boxplot(data=data_train, x = 'SaleType', y = 'SalePrice', ax=axis[1,1])
axis[1,1].set_title('SaleType Vs Sale Price')

plt.tight_layout()
```

So, what did we get from the above graph. If the sale type is new then the price of the home increase same can be said for the partial sale condition. I am not too sure about the Condition1 and Condition2 as the attributes are not too familiar.
