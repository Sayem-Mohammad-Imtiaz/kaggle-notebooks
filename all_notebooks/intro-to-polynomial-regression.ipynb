This section deals with Polynomial Regression. Begin by the usual importing of essential packages.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Next, input the dataset and define the independent and dependent variables

```python
# Importing the dataset
dataset = pd.read_csv('../input/Position_Salaries.csv')
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

As there are no categorical variables to be labelencoded here. However, since we need to make as accurate a prediction as possible, and that we only have 10 data points, we should not split into test and train dataset here.

For a comparison, we'll build a linear model and see how it fares.

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

Now we build the new matrix of features with the respective powered values using the PolynomialFeatures class

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=4)
X_poly = poly_reg.fit_transform(X)
```

Now we can build the polynomial regression model on this by fitting X_poly to a lin reg.

```python
poly_lin_reg = LinearRegression()
poly_lin_reg.fit(X_poly, y)
```

First, let's see how well the Linear Regressor alone performed

```python
# Visualising the Linear Regression results
plt.scatter(X, y, color = 'red') # the actual data in re
plt.plot(X, lin_reg.predict(X), color = 'blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

Clearly, the linear regressor Now let's see how the linear regressor with X_poly as input performed.

```python
# Visualising the Polynomial Regression results
plt.scatter(X, y, color = 'red')
plt.plot(X, poly_lin_reg.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

Better, but still not good enough. Let's try and improve our reg by changing the degree to what works and increasing the resolution of the model.

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = 'red')
plt.plot(X_grid, poly_lin_reg.predict(poly_reg.fit_transform(X_grid)), color = 'blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```

Let's predict what the linear regressor says about predicted value of 6.5

```python
lin_reg.predict([[6.5], ])
```

Let's predict what the p regressor says about predicted value of 6.5

```python
poly_lin_reg.predict(poly_reg.fit_transform([[6.5], ]))
```

Clearly, the polynomial regression gives the more accurate result of 160000 that we were expecting from the start. :)
