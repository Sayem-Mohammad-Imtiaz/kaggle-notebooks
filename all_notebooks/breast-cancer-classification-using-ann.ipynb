```python
df = df.drop(['Unnamed: 32', 'id'], axis = 1)
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['diagnosis'] = label_encoder.fit_transform(df['diagnosis'])
```

```python
X = df.iloc[:,1:].values
y = df.iloc[:,0].values
```

```python
from sklearn.preprocessing import StandardScaler

std_scaler = StandardScaler()
X = std_scaler.fit_transform(X)
```

```python
import tensorflow as tf
from keras import Sequential
from keras.layers import Dense, Dropout
```

```python
def build_model():
  tf.keras.backend.clear_session()


  basic_model = Sequential([
                            Dense(32, activation='relu', input_shape=(30,),
                                  kernel_initializer='uniform'),
                            Dense(16, activation = 'relu',
                                  kernel_initializer='uniform'),
                            Dense(16, activation = 'relu',
                                  kernel_initializer='uniform'),
                            Dense(4, activation = 'relu',
                                  kernel_initializer='uniform'),
                            Dense(1, activation='sigmoid')
  ])

  basic_model.compile(optimizer='adam', loss= 'binary_crossentropy',
                    metrics =['acc'])
  
  return basic_model
```

```python
model = build_model()
model_history = model.fit(X, y, 
                batch_size=512, epochs=100, validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
def plot_graph(model_history, epochs):
  epochs = range(1, epochs+1)
  fig, (ax1, ax2) = plt.subplots(1, 2)
  fig.set_size_inches(20, 5)

  original_loss = model_history.history['loss']
  original_val_loss = model_history.history['val_loss']
  original_acc = model_history.history['acc']
  original_val_acc = model_history.history['val_acc']

  ax1.plot(epochs, original_loss, 'bo', label='Train Loss')
  ax1.plot(epochs, original_val_loss, 'ro', label='Loss')
  ax1.set_title('Loss Plot')
  ax1.set(xlabel='Epochs', ylabel='Validation Loss')
  ax1.legend()
  ax2.plot(epochs, original_acc, 'bo', label='Train Acc')
  ax2.plot(epochs, original_val_acc, 'ro', label='Validation Acc')
  ax2.set_title('Accuracy Plot')
  ax2.set(xlabel='Epochs', ylabel='Accuracy')
  ax2.legend()
  plt.show()

plot_graph(model_history, 100)
```
