```python
from tensorflow import keras 
import numpy as np
```

```python
data=keras.datasets.imdb
```

```python
(train_data, train_labels),(test_data, test_labels)=data.load_data(num_words=10000)
```

```python
train_data[:3]
```

```python
train_labels[:10]
```

```python
word_index=data.get_word_index()
```

```python
word_index
```

```python
my_word_index = {}
for k, v in word_index.items():
  my_word_index[k]= v+3

my_word_index["<PAD>"] = 0
my_word_index["<start>"] = 1
my_word_index["<DOV"] = 2
```

```python
reverse_index = dict([( value,key) for (key, value) in my_word_index.items()])
```

```python
reverse_index[1]
```

```python
train_data2 = keras.preprocessing.sequence.pad_sequences(train_data, value=my_word_index["<PAD>"], padding="post", maxlen=300)
```

```python
test_data2 = keras.preprocessing.sequence.pad_sequences(test_data, value=my_word_index["<PAD>"], padding="post", maxlen=300)
```

```python
def decode(text):
  txt=""
  for i in text:
    word = reverse_index.get(i, "<DOV>")
    txt += " "+ word
  return txt
```

```python
print(decode(train_data2[0]))
```

```python
train_labels[0]
```

```python
model=keras.Sequential()

model.add(keras.layers.Embedding(10000, 16))
model.add(keras.layers.GlobalAveragePooling1D())
model.add(keras.layers.Dense(16 , activation='relu'))
model.add(keras.layers.Dense(1 , activation="sigmoid"))

model.compile(optimizer="adam" , loss="binary_crossentropy" , metrics=['accuracy'])
```

```python
train_data2.shape
```

```python
x_train=train_data2[:15000]
x_validation=train_data2[15000:]

y_train=train_labels[:15000]
y_validation=train_labels[15000:]
```

```python
history=model.fit(x_train, y_train, epochs=30, batch_size=128,validation_data=(x_validation,y_validation) ,verbose=1 )
```

```python
evaluation=model.evaluate(test_data2, test_labels)
```
