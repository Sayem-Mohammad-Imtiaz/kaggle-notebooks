```python
# import packages
import numpy as np
import pandas as pd
```

```python
# Check out the data
data = pd.read_csv("../input/road-weather-information-stations.csv")
data = data[['StationName', 'StationLocation']]
data.head(10)
```

```python
# Subset the data for the desired station
df = data[data['StationName'].str.contains("HarborAve")]
df.head(10)
```

```python
# Only look at unique locations
unique_loc = df.groupby('StationLocation')
```

```python
# Look at the coordinates and store them
sensor_loc = []

for name, group in unique_loc:
    print(name)
    sensor_loc.append(name)
```

```python
# Do any other sensors live at these locations?
shared_sensors = data[data['StationLocation'].isin(sensor_loc)]
station_names = shared_sensors.groupby('StationName')
```

```python
for name, group in station_names:
    print(name)
```
