```python
city_map = {'Arun Jaitley Stadium, Delhi' : 'Delhi',
'Eden Gardens, Kolkata' : 'Kolkata',
'M. Chinnaswamy Stadium, Bengaluru' : 'Bengaluru',
'MA Chidambaram Stadium, Chepauk, Chennai' : 'Chennai',
'Narendra Modi Stadium, Ahmedabad' : 'Ahmedabad', 
'Wankhede Stadium, Mumbai' : 'Mumbai'}

home_team_map = {'Arun Jaitley Stadium, Delhi' : 'DC',
'Eden Gardens, Kolkata' : 'KKR',
'M. Chinnaswamy Stadium, Bengaluru' : 'RCB',
'MA Chidambaram Stadium, Chepauk, Chennai' : 'CSK',
'Wankhede Stadium, Mumbai' : 'MI'}

venue_colors = {'Mumbai' : '#0722ab',
'Chennai' : '#fdb913',
'Delhi' : '#004c93',
'Kolkata' : '#3d2256',
'Ahmedabad' : '#545252',
'Bengaluru' : '#ef4022'}

fix21['city'] = fix21['venue'].apply(lambda x : city_map[x])

def autopct_format(values):
    def my_format(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{v:d}'.format(v=val)
    return my_format

#colors
teams = sorted(fix21.loc[~fix21['home'].isna()]['home'].unique())
#explsion
explode = (0.05,0.05,0.05,0.05)

fig1, ax1 = plt.subplots(4,2, figsize=(12, 12), facecolor='w', edgecolor='b')
fig1.subplots_adjust(hspace = .5, wspace=.001)
axs = ax1.ravel()
for i in range(8):
    df_pie = fix21.loc[(fix21['home'] ==teams[i]) |(fix21['away'] ==teams[i])].groupby(['city'])['round'].count().reset_index()
    # Pie chart
    labels = df_pie['city']
    sizes = df_pie['round']
    patches, texts, autotexts = axs[i].pie(sizes, 
            colors = [venue_colors[key] for key in labels], 
            labels=labels, 
            autopct=autopct_format(sizes), 
            startangle=90, 
            pctdistance=0.85, 
            explode = explode,
           textprops={'fontsize': 14,
                     'fontfamily':'Computer Modern'
                      })
    
    [text.set_color('#4a4b52') for text in texts]
    [autotext.set_color('white') for autotext in autotexts]
    [autotext.set_weight('bold') for autotext in autotexts]
    #draw circle
    centre_circle = plt.Circle((0,0),0.70,fc='white')
    axs[i].add_artist(centre_circle)
    # Equal aspect ratio ensures that pie is drawn as a circle
    axs[i].axis('equal')  
    axs[i].set_title(teams[i],fontsize = 18)
plt.tight_layout(pad=3.0)
plt.subplots_adjust(top=0.91)
plt.suptitle('Number of games in each city by teams',fontsize = 20)
plt.show()
```

```python
def draw_table(fix21, title):
    fix21['date'] = pd.to_datetime(fix21['date'], format='%Y-%m-%d %H:%M:%S')
    rowEvenColor = 'lightgrey'
    rowOddColor = 'white'

    color = [[rowOddColor,rowEvenColor,rowOddColor,rowEvenColor]*len(fix21.index)]

    fig = go.Figure(data=[go.Table(
        columnwidth = [1.25,1.5,1.5,2,0.5],
        header=dict(values=['<b>Date</b>',
                            '<b>Home Team</b>', 
                            '<b>Away Team</b>', 
                            '<b>Venue</b>',
                           '<b>Round</b>'],
                    fill_color='#545252',
                    font=dict(color='white',size=12),
                    line_color='darkslategray',
                    align=['left','center']),
        cells=dict(values=[fix21['date'].dt.strftime('%d/%m/%Y %H:%M'), 
                           fix21['home_team'], 
                           fix21['away_team'], 
                           fix21['venue'],
                           fix21['round']],
                   fill_color=color,
                   line_color='darkslategray',
                   align=['left','center'],
                  font=dict(color='darkslategray',size=13)
                  )
        )
    ])
    fig.update_layout(title=dict(text=title,
                                                       x=0.5,
                                                       font_color='black'),
                      margin=dict(r=20, l=20)
                     )


    iplot(fig)
```

### Away from home performance
[back to top](#0.0)

> Below chart shows the number of wins by each team across venues for 2021 season. White color indicates that the venue is home ground for the team, hence its excluded from the chart since there is no home ground advantage this season

```python

def get_loser(row):
    if row['home_team'] == row['winner']:
        return row['away_team']
    else:
        return row['home_team']

def update_venue(row):
    if row['venue_name'] == 'M.Chinnaswamy Stadium, Bengaluru':
        return "M. Chinnaswamy Stadium, Bengaluru"
    else:
        return row['venue_name']

def calculate_total_games(row):
    total_games = all_summary.loc[(all_summary['venue_name'] == row['venue_name']) & ((all_summary['home_team'] == row['winner']) |  (all_summary['away_team'] == row['winner']))].shape[0]
    return total_games
    
all_summary['loser'] = all_summary.apply(get_loser, axis=1)
all_summary['venue_name'] = all_summary.apply(update_venue, axis=1)

win_df = pd.DataFrame({'count' : all_summary.groupby(['venue_name','winner'])['winner'].count()}).reset_index()
win_df['total_games'] = win_df.apply(calculate_total_games, axis=1)
win_df['win_percent'] = round(win_df['count'] / win_df['total_games'],2)*100
win_df = win_df.loc[win_df['winner'].isin(['MI', 'CSK', 'SRH', 'RR', 'KKR', 'KXIP', 'RCB', 'DC'])]

for key, value in home_team_map.items():
    win_df = win_df.loc[~((win_df['venue_name'] == key) & (win_df['winner'] == value))]
win_df = win_df.loc[win_df['venue_name'].isin(home_team_map.keys())] 
win_df = win_df.reset_index(drop=True)

import plotly.graph_objects as go

fig = go.Figure(data=go.Heatmap(name='Win-Stats',
                   z=win_df['count'],
                   x=win_df['winner'],
                   y=win_df['venue_name'],
                   hoverongaps = True,
                   colorscale = 'viridis_r',
                    hovertemplate = 'Venue : %{y}<br>Team : %{x}<br><b>Wins :%{z}</b>'))
fig.update_layout(xaxis= { 'tickfont' : {'size':18},  "zeroline":False, "showgrid":False,
                         "fixedrange": False,
                                                    "tickangle" : 45
                        }, 
                   title = { 'text' : 'Number of wins',
                          'x':0.5,
                        'xanchor': 'center',
                        'yanchor': 'top'
                        },
                  #title_font_family = 'Computer Modern',
                 yaxis={'title':None,"zeroline":False, "showgrid":False,
                         "fixedrange": False},
                 font=dict(family='Computer Modern',size=15, color="#00040a"),
                 showlegend=True,
                plot_bgcolor="#f5f3e9",
                margin={"r":20, "l":30}
                 )
iplot(fig)
```

```python
def plot_batting_chart(title, column, label):
    runs = batting.groupby(['fullName','current_innings'])[column].sum().sort_values(ascending=False).reset_index().head(10)
    fig = px.bar(runs, x='fullName', y=column,
             hover_data=['fullName', column], color='current_innings',
             labels={column: label},
             color_discrete_map=team_colors,
             text=column)
    fig.update_layout(xaxis={'categoryorder':'array', 'categoryarray': runs['fullName'],'title' : 'Batsman', 'showgrid':False},
                      yaxis={'showgrid':False},
                     title = title)
    fig.show()
```

```python
def plot_team_batting_chart(title, column, label):
    df = batting.groupby(['current_innings'])[column].sum().reset_index()
    df = df.sort_values(column, ascending=False)
    fig = px.bar(df, x='current_innings', y=column,
             hover_data=['current_innings', column], color='current_innings',
             labels={column: label},
             color_discrete_map=team_colors,
             text=column)
    fig.update_layout(xaxis={'categoryorder':'array', 'categoryarray': df['current_innings'],'title' : 'Team','showgrid' :False},
                      yaxis=dict(title='Total',showgrid=False,zeroline=False),
                     title = title)
    fig.show()
```

> Its always a good idea to look further than just taking average of an feature. There could be outliers. In this case, the following boxplot chart reveals that there are few outliers. Best example in case of CSK, there is a strike rate of 600. Well this is none other than [Ravindra Jadeja](https://en.wikipedia.org/wiki/Ravindra_Jadeja) (a.k.a Jaddu for CSK fans) hitting a six off the last ball of CSK vs KKR innings and remaining not out. So 6 runs in 1 ball, hence the Strike Rate of 600. That truely is an outlier (so does Jaddu among the cricketing talents !). There are few other outliers as well. So its better to look into median or IQR of the strike rate to judge overall batting performance of teams
