```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Reading the data and putting in Independent and Dependent Variables

```python
dataset = pd.read_csv('../input/linear.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

Checking the nature of the columns by seeing the first few rows

```python
dataset.head()
```

Getting some information about the dataset

```python
dataset.info()
dataset.describe()
dataset.columns
```

Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

Importing the Linear Regression Model and fitting it on the training data

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
Predicting the data based on the model trained in the previous step
```

```python
y_pred = regressor.predict(X_test)
plt.scatter(y_test,y_pred) # Scatter plot to compare test data vs predicted data
plt.xlabel('Y Test')
plt.ylabel('Y Predicted')
```

Residual Histogram

```python
sns.distplot((y_test-y_pred),bins=5);	
```

Evaluating the Model

```python
print(regressor.intercept_)
```

Regression Evaluation Metrics

```python
from sklearn import metrics  # Importing the metrics module
print('Mean Absolute Error (MAE):', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Square Error (MSE):', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Square Error (RMSE):', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

 Visualizing the Training Set Results

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train,regressor.predict(X_train), color ='blue')
plt.title('Age Vs Quality (Training set)')
plt.xlabel('Age')
plt.ylabel('Quality')

plt.show()
```

 Visualizing the Test Set results

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train,regressor.predict(X_train), color ='blue')
plt.title('Age Vs Quality (Test set)')
plt.xlabel('Age')
plt.ylabel('Quality')

plt.show()
```
