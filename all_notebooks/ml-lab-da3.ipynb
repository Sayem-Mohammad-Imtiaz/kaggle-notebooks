```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
df = pd.read_csv('../input/hr-analytics/HR_comma_sep.csv')
df.head()
```

```python
dflr = df[['satisfaction_level','average_montly_hours','promotion_last_5years','salary']]
dflr.head()
```

```python
salary_dummies = pd.get_dummies(dflr.salary,prefix='salary')
df_dummies = pd.concat([dflr,salary_dummies],axis='columns')
df_dummies.drop('salary',axis='columns',inplace=True)
df_dummies.head()
```

```python
X = df_dummies
Y = df.left
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
model = LogisticRegression()
model.fit(X_train,Y_train)
```

```python
pred=model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```

```python
from sklearn.metrics import plot_confusion_matrix
matrix = plot_confusion_matrix(model, X_test, Y_test,cmap=plt.cm.Blues,normalize='true')
plt.title('Confusion matrix for our classifier')
plt.show(matrix)
plt.show()
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
#calculating precision and reall
precision = precision_score(Y_test,pred)
recall = recall_score(Y_test,pred)
f1score = f1_score(Y_test,pred)
print('Precision: ',precision)
print('Recall: ',recall)
print('F1-Score: ',f1score)
```
