# Simple Analysis - True Car Listings 2017


## Import Packages

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.linear_model import LinearRegression
```

## Load the Data

```python
raw_data = pd.read_csv('../input/true-car-listings-2017-project/true_car_project_full.csv')
raw_data.head()
```

## Preparing for Analysis

### Create a copy of the data

```python
data = raw_data.copy()
```

### Remove the Id and Vin fields (not necessary)

```python
data = data.drop(['Id'], axis = 1)
data = data.drop(['Vin'], axis = 1)
```

```python
data.head()
```

### Descriptive Statistics

```python
data.describe()
```

```python
# Median is more helpful than average/mean with a million rows of data
data.median()
```

### The data field will be too wide, so set parameters: only Ford make, Price <  $100,000, Mileage < 200,000, and no older than 15 years.

```python
data_ford = data.copy()
```

```python
data_ford.drop(data_ford.index[data_ford['Make'] != 'Ford'], inplace = True)
data_ford.drop(data_ford.index[data_ford['Price'] > 100000], inplace = True)
data_ford.drop(data_ford.index[data_ford['Mileage'] > 200000], inplace = True)
data_ford.drop(data_ford.index[data_ford['Year'] < 2006], inplace = True)
```

```python
# Declare mileage as the independent variable and price as the dependent variable.
x = data_ford['Mileage']
y = data_ford['Price']
```

```python
# Set the ScatterPlot
plt.scatter(x,y)
plt.xlabel('Ford Mileage',fontsize=20)
plt.ylabel('Ford Price',fontsize=20)
plt.show()
```
