```python
first_q = (df['count'].max() + mean)/2
lower_q = (df['count'].min() + mean)/2
first_q, lower_q
```

```python
x_label = []
y_label = []
range_dim =np.arange(10000, len(df),10000, dtype=None)
start = 0
batch = 0
for rn in range_dim:
    df_range = df[(df['index'] > start) & (df['index'] < rn)]
    start = rn
    batch = batch + 1
    x_label.append(batch)
    y_label.append((sum(df_range['count'])/len(df['count'])))
    
```

```python
import matplotlib.pyplot as plt

 
names = list(x_label)
values = list(y_label)
 
plt.plot(names, values)
plt.xlabel("Batch") # 1 batch with 10000 words
plt.ylabel("Word count")
```
