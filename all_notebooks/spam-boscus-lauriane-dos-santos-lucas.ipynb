```python
from nltk.tokenize import word_tokenize
import string
from nltk.corpus import stopwords
stop=set(stopwords.words('english'))
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
sms['Message']=sms['Message'].str.lower()
```

```python
def remove_punct(text):
    text_tok = word_tokenize(text)
    L = []
    for i in text_tok:
        if not i in string.punctuation:
            L.append(i)
            
    resultat = " ".join(L)
    return resultat
```

```python
sms['Message']=sms.Message.apply(remove_punct)
```

```python
def remove_stopword(text):
    text_tok = word_tokenize(text)
    l = []
    for a in text_tok:
        if not a in stop:
            l.append(a)
            
    resultat = " ".join(l)
    return resultat
```

```python
sms['Message']=sms.Message.apply(remove_stopword)
```

```python
def lemm(text):
    text_tok = word_tokenize(text)
    l = []
    for i in text_tok:
        l.append(lemmatizer.lemmatize(i))
    resultat = " ".join(l)
       
    return resultat
```

```python
sms['Message']=sms.Message.apply(lemm)
```
