# Matplotlib
### Datenvisualisierung

```python
# importiere Matplotlib.pyplot
import matplotlib.pyplot as plt

# Generiere einen Test-Datensatz
import numpy as np
from random import random

x = np.arange(10)

y = np.array([round(2*i + 3*random(), 1) for i in x])

print("x: \n", x)
print("y: \n", y)
```

Unterschiedliche Plot-Typen

```python
#Einfacher Plot
plt.plot(x, y)

plt.show()
```

```python
# Scatter-Plot
plt.scatter(x, y)

plt.show()
```

```python
plt.bar(x, y)

plt.show()
```

```python
#Histogramm

# Hier wird ein neuer Datensattz generiert, der mehr Sinn macht im Umgang mit Histogrammen (eine Verteilung)
from random import normalvariate

x = np.array([normalvariate(7,1) for i in range(1000)])

# plotten eines Histogramms
plt.hist(x, bins = 10)

plt.show()
```

```python
# Zurück zum alten selbstgenerierten Datensatz (mit einer Erweiterung durch y2)
x = np.arange(10)

y1 = np.array([round(2*i + 3*random(), 1) for i in x])
y2 = np.array([round(2.3*i + 2*random(), 1) for i in x])
```

### Im Folgenden werden die einzelnen Schritte Zelle für Zelle für ein vollständiges Diagramm erweitert. Dies geschieht eigentlich alles in einer Zelle, soll ich aber dazu dienen die einzelnen Schritte besser nachzuvollziehen.

#### Neue Schritte sind mit einem # markiert.

```python
# In einer Figur lassen sich mehrere plots realisieren
plt.plot(x, y1)
plt.plot(x, y2)

plt.show()
```

```python
# Hinzufügen eines Titels und Achsenbeschriftung
plt.plot(x, y1)
plt.plot(x, y2)

plt.title("Der Titel meines Plots") #
plt.xlabel("x-Achse") #
plt.ylabel("y-Achse") #


plt.show()
```

```python
# Die Farben der Datenpunkte/Linien, als auch ihr Style lassen sich anpassen.
plt.plot(x, y1, color = "green", linestyle = "--") #
plt.plot(x, y2, color = "red") #

plt.title("Der Titel meines Plots")
plt.xlabel("x-Achse")
plt.ylabel("y-Achse")

plt.show()
```

```python
# Einfügen eines Legende und passender Label
plt.plot(x, y1, color = "green", linestyle = "--", label = "y1") #
plt.plot(x, y2, color = "red", label = "y2") #

plt.title("Der Titel meines Plots")
plt.xlabel("x-Achse")
plt.ylabel("y-Achse")
plt.legend(loc = "upper left") #

plt.show()
```

```python
# Es lassen sich sonstige Optimierungen an der Figur realisieren, zB Skalierung,....
plt.plot(x, y1, color = "green", linestyle = "--", label = "y1")
plt.plot(x, y2, color = "red", label = "y2")

plt.title("Der Titel meines Plots")
plt.xlabel("x-Achse")
plt.ylabel("y-Achse")
plt.legend(loc = "upper left")

plt.xlim(0, 8) #
plt.xticks(np.arange(0, 9, 2), labels = range(2010, 2019, 2), rotation = 90) #

plt.show()
```

# Aufgabe 2

1)
Erstelle einen einfachen Plot, der die Lohnentwicklung über die Jahre von Manager*innen (f und m) illustriert

2)
Erstelle einen weiteren Plot, der die Lohndifferenzen zwischen female und male darstellt. Hierfür wollen wir die Lohndifferenz für Manager und für Maschinenoperateur*innen in einem Plot haben

** Sorge dafür dass deine Plots alle wichtigen Informationen enthalten:**
* Titel
* Achsenbeschriftung
* Legende und Label

```python
# Lösung 1

plt.plot(data_red["year"], data_red["manager_f"], label = "ManagerInnen (w)")
plt.plot(data_red["year"], data_red["manager_m"], label = "ManagerInnen (m)")

# Hier habe ich noch zusätzlich den Durchschnittslohn aller (also m und w) hinzugefügt,
# denn hier können wir auch anhand der Nähe zu gewissen plots herleiten,
# dass eine Gruppe zahlmäßig stärker vertreten ist
plt.plot(data_red["year"], data["personmanagers"], label = "ManagerInnen (total)")


plt.title("Lohnentwicklung w vs m")
plt.xlabel("Jahr")
plt.ylabel("Stundenlohn in US $")
plt.legend()

plt.style.use("seaborn")

plt.show()
```

```python
# Lösung 2
# alternativ kann man Lohndifferenz auch vorher berechnen
plt.plot(data_red["year"], data_red["manager_m"] - data_red["manager_f"], label = "ManagerInnen")
plt.plot(data_red["year"], data_red["operator_m"] - data_red["operator_f"], label = "OperatorInnen")

# Eine Hilfslinie, die uns die Nulllinie generiert
plt.plot(data_red["year"], np.zeros(len(data_red["year"])), color = "red", linestyle = "--")

plt.title("Lohndifferenz Entwicklung")
plt.ylabel("Differenz in US $")
plt.xlabel("Jahre")
plt.legend()
plt.xlim(2004, 2017)

plt.show()
```
