# Perth maximum and minimum temperature prediction
### Purpose: To predict the maximum and minimum temperature in Perth with the date
### Data description: Data include two parts: dates(years, months, days), maximum temperature, minimum tamperature and waterfall(ignored)


### Hypothesis: I believe that randomforest will be the best model because the temperature model should be periodic depending on the days, years and especially months.

# Setting the code
### Import the essential functions for reading csv file, models, validation and ploting the graph.

### After gathering data, dividing them into two parts:features(date) and goals(minimum and maximum temperature)

```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import datasets, linear_model
```

# Gathering the data
### Read file, drop missed values and seperate them into two parts: the date(required data for prediction) and the temperature(values needed to predicted), drop the missing values and waterfall column.

```python
perth_file_path = '../input/perth-temperatures-and-rainfall/PerthTemperatures.csv'
ptemp_data = pd.read_csv(perth_file_path)
ptemp_data = ptemp_data.dropna(axis=0)
ptemp_data.describe()
#y1=ptemp_data['Minimum temperature (Degree C)']
#y1=ptemp_data.Minimum_temperature_(Degree_C)
#y2=ptemp_data['Maximum temperature (Degree C)']
yfeatures=['Minimum temperature (Degree C)','Maximum temperature (Degree C)']
Y=ptemp_data[yfeatures]
pfeatures=['Year','Month','Day']
X=ptemp_data[pfeatures]
```

# Preparing the data
### Splitting the data into the training group(for building model) and the testing group(for validation)

```python
train_X, val_X, train_Y, val_Y = train_test_split(X, Y, random_state = 0)
#print(train_X+'aa\n', val_X+'aa\n', train_y1+'aa\n', val_y1+'aa\n', train_y2, val_y2)
#print(train_X+'aaa\n\n\n\n\n\n')
#print(val_X+'aaa\n\n\n')
#print(val_Y+'ddss\n\n')
print(train_Y)
print(val_Y)
```

### Visualize the data using sns

```python
plt.figure(figsize=(100,30))
sns.lineplot(data=Y)
```

### I find that the temperature changes periodicly.

# Choosing Models
### Using three models: randomforest, decisiontree and linear regression.

```python
tmodel= RandomForestRegressor(random_state=1)
tmodel.fit(train_X,train_Y)


tmodel2= DecisionTreeRegressor(random_state=2)
tmodel2.fit(train_X,train_Y)


tmodel3=linear_model.LinearRegression()
tmodel3.fit(train_X, train_Y)
```

# Making predictions

### Then, calculate the mean absolute errors and comparetofind the best model.

```python
pred=tmodel.predict(val_X)

pred2=tmodel2.predict(val_X)

pred3=tmodel3.predict(val_X)


print('prediction of randomforest: \n',pred)
print("\n\n")
print('prediction of randomforest: \n',pred2)
print("\n\n")
print('prediction of randomforest: \n',pred3)
print("\n\n")
print('actual values: \n',val_Y)
```

# Validation
### Choosing the model with least mean absolute error
### Plotting the difference between predictions and actual values.

```python
print('Mean absolute error of randomforest: \n',mean_absolute_error(val_Y, pred))

print('Mean absolute error of decision tree: \n',mean_absolute_error(val_Y, pred2))

print('Mean absolute error of linear regression: \n',mean_absolute_error(val_Y, pred3))

plt.figure(figsize=(100,30))
sns.lineplot(data=pred-val_Y)
plt.figure(figsize=(100,30))
sns.lineplot(data=pred2-val_Y)
plt.figure(figsize=(100,30))
sns.lineplot(data=pred3-val_Y)
```

# Conclusion
### Randomforest is the best model I choose
