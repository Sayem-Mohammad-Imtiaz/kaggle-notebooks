```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data0=pd.read_csv('../input/people-walking-in-kobe/kobe_people_2020.csv')
data0
```

```python
data1=data0[data0['Time']=='day time(5-19)'][data0['Area']=='center street']
data2=data0[data0['Time']=='night time(19_29)'][data0['Area']=='center street']
```

```python
data1
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=data1['Date'],y=data1['Value'],name='Day Time (5-19)'),secondary_y=False,)
fig.add_trace(go.Scatter(x=data2['Date'],y=data2['Value'],name='Night Time (19-29)'),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="Numberof Pedestrain at Center Street")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Value",secondary_y=False)
fig.show()
```

```python
weather0=pd.read_csv('../input/weather-in-kobe/kobe_weather_2020.csv')
weather0
```

```python
weather_date=list(weather0['Date'])
pedestrian_date=list(data1['Date'])

for item in weather_date:
    if item not in pedestrian_date:
        print(item+' data is missing.')
```

```python
data3=data1.merge(weather0,on='Date',how='left')
data3
```

```python
fig,ax = plt.subplots(figsize=(6,6))
ax.set_title('Pedestrian in day time vs Rain',fontsize=20)
ax.set_xlabel('Precipiation',fontsize=12)
ax.set_ylabel('Value',fontsize=12)
ax.scatter(data3['Precipitation'],data3['Value'])
```

### Number of pedestrian in day time seems to be restricted by heavy rain.
