### This pie chart represent the proportion between career levels

```python
pie_df = df.career_level.value_counts().reset_index()
pie_df.columns = ['career_level','count']
fig = px.pie(pie_df, values='count', names='career_level', title='Proportion of career_level',
             color_discrete_sequence=['#34568B', '#FFA500','#CD212A','#FA7A35',"#6A2E2A"])
fig.show()
```

### chart results
-  The most level that is requirded isExperienced

### Box plot that present career level vs average salary vs exp. years

```python
fig = px.box(df, x="career_level", y="average salary",color="exp.years")
fig.update_traces(quartilemethod="exclusive")
fig.show()
```

```python
fig = px.box(df, x="job_category1", y="average salary",color="exp.years")
fig.update_traces(quartilemethod="exclusive")
fig.show()
```

### chart results
- The least salary for a student with no experience is 3000 L.E
- The maximum salary for a 10 experience manager is 140,000
