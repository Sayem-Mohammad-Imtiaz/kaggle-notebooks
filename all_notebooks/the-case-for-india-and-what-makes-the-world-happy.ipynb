```python
import seaborn as sns
```

```python
sns.pairplot(df1,x_vars=[
 'Logged GDP per capita',
 'Social support',
 'Healthy life expectancy',
 'Freedom to make life choices',
 'Generosity',
 'Perceptions of corruption'],
    y_vars=['Ladder score'],)
```

```python
df4=df1.drop(['Country name','Regional indicator','Ladder score in Dystopia',
 'Explained by: Log GDP per capita',
 'Explained by: Social support',
 'Explained by: Healthy life expectancy',
 'Explained by: Freedom to make life choices',
 'Explained by: Generosity',
 'Explained by: Perceptions of corruption',
 'Dystopia + residual','Standard error of ladder score',
 'upperwhisker',
 'lowerwhisker'],axis=1)
df4.corr()
```

```python
sns.set_style()
plt.figure(figsize=(20,15))
sns.heatmap(df4.corr(),annot=True)
plt.show()
```

Happiness score is strongly correlated to 1.GDP per capita, 2.Life expectancy, 3.Social support. Surprising factor is that generosity is negativelty correlated to happiness score. Are all psychologists wrong? Also more the perception of corruption lesser the happiness score.

Lets look at India

```python
ddf=df2.drop(['Positive affect','Negative affect'],axis=1)
dfi=ddf.loc[ddf['Country name']=='India']
dfi.head()
```

```python
dfi.info()
```

```python
# dfi.fillna(method ='bfill')
# dfi
dfi.notnull()
```

```python
dfi.fillna(method ='bfill',inplace = True)
```

```python
#Score has declined continuously for India
plt.figure(figsize=(20,10))
sns.pointplot(x="year", y="Life Ladder",data=dfi)
sns.despine()
```

```python
#GDP per capita increasing
plt.figure(figsize=(20,10))
sns.pointplot(x="year", y="Log GDP per capita",data=dfi)
sns.despine()
```

```python
#Some decrease in Social support
plt.figure(figsize=(20,10))
sns.pointplot(x="year", y="Social support",data=dfi)
sns.despine()
```

```python
#Increased life expectancy
plt.figure(figsize=(20,10))
sns.pointplot(x="year", y="Healthy life expectancy at birth",data=dfi)
sns.despine()
```

```python
#Increased freedom to make choice
plt.figure(figsize=(20,10))
sns.pointplot(x="year", y="Freedom to make life choices",data=dfi)
sns.despine()
```

```python
#Increase in Generosity
plt.figure(figsize=(20,10))
sns.pointplot(x="year", y="Generosity",data=dfi)
sns.despine()
```

```python
#Decrease in perception of corruption
plt.figure(figsize=(20,10))
sns.pointplot(x="year", y="Perceptions of corruption",data=dfi)
sns.despine()
```

Well the outcome is indeed shocking, the jury has continuously degraded India's happiness score even tough the counrtry has almost all the positive signs for the increase or at least a stable score. Hence the general trend seems not be followed in the case of India.
