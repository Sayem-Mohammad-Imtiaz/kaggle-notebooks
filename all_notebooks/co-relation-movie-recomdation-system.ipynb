```python
moviemat=df.pivot_table(index='userId',columns='title',values='rating')
```

```python
moviemat
```

```python
JurassicPark_user_rating=moviemat['Jurassic Park (1993)']
```

```python
JurassicPark_user_rating
```

```python
simlar_to_JurassicPark=moviemat.corrwith(JurassicPark_user_rating)
```

```python
corr_JurassicPark=pd.DataFrame(simlar_to_JurassicPark,columns=["Corelation"])
corr_JurassicPark.dropna(inplace=True)
```

```python
corr_JurassicPark.sort_values("Corelation",ascending=False).head()
```

```python
corr_JurassicPark=corr_JurassicPark.join(ratings["num of rating"])
```

```python
corr_JurassicPark
```

```python
corr_JurassicPark[corr_JurassicPark["num of rating"]>100].sort_values("Corelation",ascending=False).head()
```
