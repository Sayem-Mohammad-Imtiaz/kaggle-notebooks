```python
df = df.drop(['pos_spec_dt','onset_dt'],axis =1)
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(12,4))
sns.heatmap(df.isnull(),cbar=False,cmap='coolwarm',yticklabels=False)
plt.title('Missing value in the dataset')
```

```python
df.groupby('sex')['age_group'].count()
```

```python
data_categorical = df[['current_status', 'sex', 'age_group', 'Race and ethnicity (combined)',
       'hosp_yn', 'icu_yn', 'death_yn', 'medcond_yn']]
```

```python
df_categorical_encode = pd.get_dummies(data = data_categorical, 
               columns = ['current_status', 'sex', 'age_group', 'Race and ethnicity (combined)',
       'hosp_yn', 'icu_yn', 'death_yn', 'medcond_yn'],
               drop_first =True,
               dtype='int8')
```

```python
df_categorical_encode.head(5)
```

```python
x = df_categorical_encode.iloc[:, [3, 4]].values
print(type(x))
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 4):
    km = KMeans(n_clusters = i)
    km.fit(x)
    wcss.append(km.inertia_)
    
plt.plot(range(1, 4), wcss)
plt.title('The Elbow Method', fontsize = 20)
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
km = KMeans(n_clusters = 3)
y_means = km.fit_predict(x)

plt.scatter(x[y_means == 0, 0], x[y_means == 0, 1], s = 100, c = 'pink')
plt.scatter(x[y_means == 1, 0], x[y_means == 1, 1], s = 100, c = 'yellow')
plt.scatter(x[y_means == 2, 0], x[y_means == 2, 1], s = 100, c = 'cyan')


plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s = 50, c = 'blue' , label = 'centeroid')

plt.style.use('fivethirtyeight')
plt.title('K Means Clustering', fontsize = 20)

plt.show()
```
