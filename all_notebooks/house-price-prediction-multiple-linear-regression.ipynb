> I've tried to predict the house price where the independent variables are area and number of rooms and the dependent variable is house price.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('../input/simple-house-price-prediction/house_price.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
print(regressor.coef_)
print(regressor.intercept_)
```

> This means that the base price of house is 90561.37 and the amount will be added/subracted for every 1 sq. units purchased.
> For some unit increase in area, the price will be increased by 125.853
> For some unit increase in no. of rooms, the price will decrease by 1596.93

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

> The predictions are pretty bad for this model.

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

**The R2 score is very bad for multiple linear regression.**
