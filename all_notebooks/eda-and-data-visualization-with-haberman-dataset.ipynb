```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
```

```python
df=pd.read_csv('../input/habermans-survival-data-set/haberman.csv')
df.head()
```

- No Feature/column name mentioned.

```python
df.columns=['Age', 'Op_Year_19--','Axillary_nodes', 'Survival_status' ]
df.head()
```

#### Attribute Information:
- Col 1. Age of patient at time of operation (numerical)
- Col 2. Patient's year of operation (year - 1900, numerical)
- Col 3. Number of positive axillary nodes detected (numerical)
- Col 4. Survival status (class attribute)
         - 1 = the patient survived 5 years or longer
         - 2 = the patient died within 5 year

### Objective: To find survival of patients who had undergone surgery for breast cancer

```python
# taking random sample
df.sample(2)
```

#### The task is to predict the survival of the patient i.e. we have target feature/col (Survival_status).
 - Dependent variable- col_4.
 - Independent variables- col_1, col_2, col_3.

```python
df.plot(kind='kde', x='Survival_status', y='Age', title='kde Plot')      # kind='kde' Kernel Density Estimation.
plt.show()
```

+ Most of patients age is between 30-75 (approximation).

```python
df.plot(kind='scatter', x='Survival_status', y='Age', title='Scatter Plot')
plt.show()
```

  -  Age does't seems a major factor regarding survive patient more than 5 years after operation
  -  However 80+ aged patient surely not survived more than 5 years.

```python
sb.boxplot(x='Survival_status',y='Age',data=df)
plt.show()
```

- Min age for the patients survived more than 5 years seems 30 and max age seems 76, and majority of patients age is between 42-60.
- Min age for the patients survived less than 5 years seems 35 and max age seems 85, and majority of patients age is between 45-61.
- **Numbers are not accurate, just an estimate.**

```python
#boxplot for all columns
df.iloc[:,:].boxplot(figsize=[7,8])
plt.subplots_adjust(bottom=0.50)
plt.show()
```

- Majority of patients age is between 42-61.
- Majority of operations happend in between 1960-1965.
- Axillary_nodes have so many outliers.

```python
df.plot(kind='kde', x='Survival_status', title='kde Plot')      # kind='kde' Kernel Density Estimation.
plt.show()
```

```python
#Violin Plot

sb.violinplot(x='Survival_status', y='Age', data=df, size=8)
plt.show()
```

### Checking correlation: plotting heatmap

```python
df['Age'].hist(bins=20,grid=True)
plt.show()
```

```python
#http://seaborn.pydata.org/generated/seaborn.heatmap.html

corr_mat=df.corr()
plt.figure(figsize=[22,12])
sb.heatmap(corr_mat, annot=True, cmap='RdYlBu_r')
plt.title('Correlation Matrix')
plt.show()
```

### Observation:
- 1(Red) is highly +ve correlated & 0(Blue) is highly -ve correlated.
- col_3 most +ve correlated whereas col_2 most -ve correlated.
