```python
import matplotlib.pyplot as plt

Totals = {'Feminino': df['gen_feminino'].sum()/1000000,
          "Masculino": df["gen_masculino"].sum()/1000000,
          "Não informado": df["gen_nao_informado"].sum()/1000000 }

names = list(Totals.keys())
values = list(Totals.values())
fig, ax = plt.subplots(1, 2,figsize=(9,6.5))
ax[0].bar(names[0:2], values[0:2], color='lightblue')

fig.suptitle("Eleitores por Genero [/ Milhão ]", fontsize="x-large")

ax[1].bar(names[2], values[2], color='lightgray', width=10) 
```

```python
fig, ax = plt.subplots(figsize=(12,12))
df.groupby('uf').agg('sum')[['gen_feminino',"gen_masculino"]].plot(kind='barh', ax=ax)
plt.legend(['Feminino', 'masculino'])
plt.title("Eleitores por Estado separado por genero")

plt.ylabel("Estados do Brasil")
```

```python
Data = {'16':(df['f_16'].sum()/df['total_eleitores'].sum())*100,
        '17':(df['f_17'].sum()/df['total_eleitores'].sum())*100,
        '18 a 20':(df['f_18_20'].sum()/df['total_eleitores'].sum())*100,
        '21 a 24':(df['f_21_24'].sum()/df['total_eleitores'].sum())*100,
        '25 a 34':(df['f_25_34'].sum()/df['total_eleitores'].sum())*100,
        '35 a 44':(df['f_35_44'].sum()/df['total_eleitores'].sum())*100,
        '45 a 59':(df['f_45_59'].sum()/df['total_eleitores'].sum())*100,
        '60 a 69':(df['f_60_69'].sum()/df['total_eleitores'].sum())*100,
        '70 a 79':(df['f_70_79'].sum()/df['total_eleitores'].sum())*100,
        '> 79':(df['f_sup_79'].sum()/df['total_eleitores'].sum())*100}

names = list(Data.keys())
values = list(Data.values())

fig, ax = plt.subplots(figsize=(10,10))
ax.bar(names, values, color='lightblue')

for i, v in enumerate(values):
    ax.text(i-.3 ,v + .5, "{0:.2f}%".format(v), color='k', fontweight='bold')

plt.title("Porcentagem do total de eleitores por faixa etária")
```

```python
ElderlybyState = df.groupby('uf').agg({'f_sup_79':'sum'}) 
TotalbyState = df.groupby('uf').agg({'total_eleitores':'sum'})

percentage_of_elderly_by_State = (ElderlybyState['f_sup_79'] / TotalbyState['total_eleitores'])*100

fig, ax = plt.subplots(figsize=(16,8))
percentage_of_elderly_by_State.sort_values(ascending=False).plot(kind='bar', ax=ax, color='lightblue')

plt.title("Porcentagem do total de Eleitores com idade superior a 79 anos por estado.")
plt.xlabel("Estados do Brasil")
```

```python
young = df.groupby('uf').agg({'f_16':'sum','f_17':'sum'}) 

YoungByState = young['f_17'] + young['f_16']
YoungByState_percentage = (YoungByState / TotalbyState['total_eleitores'])*100

fig, ax = plt.subplots(figsize=(16,8))
YoungByState_percentage.sort_values(ascending=False).plot(kind='bar', ax=ax, color='lightblue')

plt.title("Porcentagem do total de Eleitores com idade inferior 18 anos por estado.")
plt.xlabel("Estados do Brasil")
```

```python
"O Objetivo dessa análise é mostrar os estados mais engajamento politicamente mostrando a percentagem de eleitores que não são obrigados a votar"

nao_Obrigatorio = df.groupby('uf').agg({'f_16':'sum','f_17':'sum','f_70_79':'sum','f_sup_79':'sum'}) 
nao_Obrigatorio_perc = nao_Obrigatorio.div(TotalbyState['total_eleitores'], axis=0)*100

nao_Obrigatorio_perc['total']=nao_Obrigatorio_perc.sum(axis=1)

fig, ax = plt.subplots(figsize=(16,8))
nao_Obrigatorio_perc =nao_Obrigatorio_perc.sort_values(by='total',ascending=False)
nao_Obrigatorio_perc[['total']].plot(kind='area',ax=ax, color='lightgray')
nao_Obrigatorio_perc[['f_16','f_17','f_70_79','f_sup_79']].plot(kind='bar',ax=ax)

plt.title("Percentagem de eleitores em faixa etaria de voto facultativo por estado")
plt.xlabel("Estados do Brasil")
```
