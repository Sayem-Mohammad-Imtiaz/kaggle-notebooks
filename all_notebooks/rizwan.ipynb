```python
### Removing unanted columns and preprocessing column headers ###
import pandas as pd
df = pd.read_csv('../input/up-crime-data-year-2019/MR data - Compiled Data Set.csv')
df.drop(['Text','Title'],axis = 1,inplace = True)
df.columns = [i for i in range(0,len(df.columns))]
df.head(20)
```

```python
### Preprocessing the data ###
df[11] = 0
df[12] = 0
df.fillna(0,inplace = True)
for i in range(0,len(df)):
    if len(str(df.iloc[i,8]).split(','))>2:
        print(df.iloc[i,8])
```

```python
dic = {0:0,'Property Disputes':1,'Family Dispute':2,'Petty Quarrels':3,'Money Disputes':4,'Personal Vendetta':5,'Love Affairs':6,'Casteism':7 ,'Unknown reasons':8}
df[1] = df[1].apply(lambda x : dic[x])
df.head(20)
```

```python
for i in range (0,len(df)):
    if ',' in str(df.iloc[i,8]):
        cl = df.iloc[i,8].split(',')
        df.iloc[i,8] = int(cl[0])
        df.iloc[i,11] = int(cl[1])
        if len(cl)>2:
            df.iloc[i,12] = int(cl[2])
df.head(20)
```

```python
df.groupby(0).sum()
### From column 9,10 we can see that the number of females victims are most in Lucknow and least in Kanpur ###
### From column 2,3 we can infer that most male victims and least child victims are in Lucknow ###
### There is no kidnapping of child in Kanpur,and no kidnapping of female in Lucknow and also that Kanpur is least prone to kidnapping ###
### It can also be infered that mostly kidnapping of a child takes place in UP(column 5) ### 
### Least number of incident took place in Kanpur and most in Lucknow ###
```

```python
import matplotlib.pyplot as plt
import numpy as np
D = df.groupby(1).sum()
D
```

```python
width = 0.25
index = np.arange(9)

plt.bar(index,D[2],width,color = 'blue',label = 'Number of child involved')
plt.bar(index+width,D[3],width,color = 'green',label = 'Number of males involved')
plt.bar(index+width+width,D[4],width,color = 'red',label = 'Number of females involved')
plt.title('Murder Victims')

plt.xlabel("Murder reason")
plt.ylabel("NUmber of victims")

plt.xticks(index+0.375,dic.keys(),rotation = 90)
plt.legend(loc='best')
plt.show()

### For some unknown reason there were a large number of males and females murdered ###
### No child was murdered because of property dispute,love affair,Personal Vendetta,Casteism ###
### No female was murdered because of Money disputes ###
### Males were murdered for every reason ###
### Very few people were murdered beacuse of Casteism ###
### Majority of murders were because of Love affairs or some unknown reason ###
```
