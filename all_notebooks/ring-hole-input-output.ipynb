```python
cols = ["TEMP","TILT_ANGLE","SILICON_CONTENT","COPPER_CONTENT","NICKEL_CONTENT","MAGNESIUM_CONTENT"]
X = df[cols].values
y = df[["PERCENTAGE_SCRAP"]].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X[:,1:] = sc.fit_transform(X[:,1:])
#y[:,1:] = sc.fit_transform(y[:,1:])
X
```

```python
print(X.shape,y.shape)
```

# **Train Test Split**

```python
from sklearn.model_selection import train_test_split

X_train,X_valid,y_train,y_valid = train_test_split(X,y,test_size=0.2,random_state=7) #20% is for validation

print(X_train.shape,X_valid.shape)
```

# Linear Regression

```python
# baseline model
from sklearn.linear_model import LinearRegression #importing lin reg model from sklearn package 
model1 = LinearRegression() # initialize the model
model1.fit(X_train,y_train) # fitting the model
pred = model1.predict(X_valid) #making the prdictions
from sklearn.metrics import r2_score # rsquare value
score1= r2_score(y_valid,pred)
print(score1)
```

# **Decision Tree**

```python
from sklearn.tree import DecisionTreeRegressor #importing pakage

dec_tree_model = DecisionTreeRegressor(random_state=7) # initiating model with any hyperparameter tuning

dec_tree_model.fit(X_train,y_train)  #fitting model

y_pred_2 = dec_tree_model.predict(X_valid) #making predictions on valdation set

score2= r2_score(y_valid,y_pred_2)
print(score2)
```

# **Random Forest**

```python
from sklearn.ensemble import RandomForestRegressor

random_forest_model = RandomForestRegressor(random_state=7)

random_forest_model.fit(X_train,y_train)

y_pred_3 = random_forest_model.predict(X_valid)

score3= r2_score(y_valid,y_pred_3)
print(score3)
```
