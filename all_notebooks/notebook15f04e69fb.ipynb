```python
#i want to know the counts of crimes in every year 
import seaborn as sns
import matplotlib.pyplot as plt
crime_group=data['Incident'].groupby(data['Year'])
crime_count_by_year=crime_group.sum()
print(crime_count_by_year)
```

```python
# then we can know the crimical counts of each year .
crime_count_dataframe=pd.DataFrame(crime_count_by_year)
crime_count_dataframe.plot()
plt.show()
```
