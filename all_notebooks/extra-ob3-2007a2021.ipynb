```python
def Extracao_2007a2019 (ano):
    ano = str(ano)
    source_file_name = '../input/bmfbovespas-time-series-19862019/COTAHIST_A'+ano+'.txt'
    print('Ano: {ano}   -   File: {file} '.format(ano=ano, file=source_file_name))

    dept_columns = ['TIPOREG', 'DATA', 'CODBDI', 'CODNEG', 'TPMERC', 'NOMRES', 'ESPECI', 'PRAZOT', 'MODREF', 'PREABE', 'PREMAX', 'PREMIN', 'PREMED', 'PREULT', 'PREOFC', 'PREOFV', 'TOTNEG', 'QUATOT', 'VOLTOT', 'PREEXE', 'INDOPC', 'DATVEN', 'FATCOT', 'PTOEXE', 'CODISI', 'DISMES']
    dept_widths = [2, 8, 2, 12, 3, 12, 10, 3, 4, 13, 13, 13, 13, 13, 13, 13, 5, 18, 18, 13, 1, 8, 7, 9, 12, 3]

    IBOVano_df = pd.read_fwf(source_file_name, widths=dept_widths, header=None, names=dept_columns)
    IBOVano_df = IBOVano_df[IBOVano_df['TIPOREG'] == 1]
    
    #display ('Ano: %', %ano, 'IBOVano_df.shape:', IBOVano_df.shape)
    print('Ano: {ano}   -   Shape: {shape} '.format(ano=ano, shape=IBOVano_df.shape))
    
    return(IBOVano_df)
    
```

```python
B3_2007A2021_df = pd.DataFrame()

for ano in range(2007, 2020, 1):
    TEMP_df = pd.DataFrame()
    TEMP_df = (Extracao_2007a2019(ano))
    B3_2007A2021_df = B3_2007A2021_df.append(TEMP_df, ignore_index=True)
    #display ('+ Ano: ', ano, 'B3_2007A2021_df.shape:', B3_2007A2021_df.shape)    
    print('+ ACUMULADO - Ano: {ano}   -   Shape: {shape} '.format(ano=ano, shape=B3_2007A2021_df.shape))
```

```python
# Processando 2020 e 2021 que estão em outros Data Frame

#2020:
ano = 2020
print('Ano: {ano}   -   Shape: {shape} '.format(ano=ano, shape=IBOV2020_df.shape))
B3_2007A2021_df = B3_2007A2021_df.append(IBOV2020_df, ignore_index=True)
print('+ ACUMULADO - Ano: {ano}   -   Shape: {shape} '.format(ano=ano, shape=B3_2007A2021_df.shape))

#2021:
ano = 2021
print('Ano: {ano}   -   Shape: {shape} '.format(ano=ano, shape=IBOV2021_df.shape))
B3_2007A2021_df = B3_2007A2021_df.append(IBOV2021_df, ignore_index=True)
print('+ ACUMULADO - Ano: {ano}   -   Shape: {shape} '.format(ano=ano, shape=B3_2007A2021_df.shape))
```

```python
#BBDC3_df = BBDC3_df.set_index('DATA')
#ACOESFECH_df['BBDC3'] = BBDC3_df['PREULT']

def CriarTodasAcoes(IBOV_df, ACOES, ACOESFECH_df):

    Temp = pd.DataFrame()
    Temp = IBOV_df[IBOV_df['CODNEG'] == ACOES]
    Temp.reset_index(drop=True, inplace=True)
    Temp = Temp.set_index('DATA')
    ACOESFECH_df[ACOES] = Temp['PREULT']
    return (ACOESFECH_df)
```

```python
# ACOES_df, ACOES, ACOESFECH_df

print (ACOESFECH_df)

for ACAO in ["EMBR3"]:
    ACOESFECH_df = CriarTodasAcoes(IBOV_df, ACAO, ACOESFECH_df)

print (ACOESFECH_df)
```
