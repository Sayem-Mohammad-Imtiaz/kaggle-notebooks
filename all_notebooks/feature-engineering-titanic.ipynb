### **Mean, median and mode imputation**

It should be used when the data is missing completely at random(MCAR)

```python
df=pd.read_csv("../input/titanicdataset-traincsv/train.csv",usecols=['Age','Fare','Survived'])
```

```python
df.isnull().mean()
```

```python
def impute_median(df,variable,median):
    df[variable+'med']=df[variable].fillna(median)
```

```python
med=df.Age.median()
med
```

```python
impute_median(df,'Age',med)
df[df['Age'].isnull()]
```

```python
print(df['Age'].std())
print(df['Agemed'].std())
```

```python
fig = plt.figure() 
ax = fig.add_subplot(111) 
df['Age'].plot(kind='kde', ax=ax) 
df.Agemed.plot(kind='kde', ax=ax, color='red') 
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines,labels,loc='best')
```

**Adv of Mean/median imputation**
* Robust to outliers
* Faster way to obtain the complete dataset

**Disadv:**
* Change or distortion in  original variance
