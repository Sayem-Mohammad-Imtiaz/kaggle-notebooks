```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.preprocessing import Normalizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
df = pd.read_csv('../input/iris-flower-dataset/IRIS.csv')
df.head()
```

```python
species = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(species)
df['species'] = le.fit_transform(df['species'].astype(str))
```

```python
df.head()
```

```python
sns.countplot(df['species'])
```

```python
hist = df.hist(figsize = (15,15))
```

```python
#split the data
x = df.drop('species', axis = 1)
y = df['species']
```

```python
scaler = Normalizer()
X = scaler.fit_transform(x)
X = pd.DataFrame(X, columns = x.columns)
X.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size = 0.3, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
KnnModel = KNeighborsClassifier(n_neighbors = 13)
KnnModel.fit(x_train, y_train)
y_pred = KnnModel.predict(x_test)

print("Accuracy score using knn is  : ", accuracy_score(y_test, y_pred))
print("Accuracy score using knn is  : ", confusion_matrix(y_test, y_pred))
print("Accuracy score using knn is  : ", classification_report(y_test, y_pred))
```

```python
TreeModel = DecisionTreeClassifier()
TreeModel.fit(x_train, y_train)
y_prediction = TreeModel.predict(x_test)

print("Accuracy score using Tree is  : ", accuracy_score(y_test, y_prediction))
print("Accuracy score using Tree is  : ", confusion_matrix(y_test, y_prediction))
print("Accuracy score using Tree is  : ", classification_report(y_test, y_prediction))
```

```python
ensembleModel = RandomForestClassifier()
ensembleModel.fit(x_train, y_train)
y_predictions = ensembleModel.predict(x_test)

print("Accuracy score using ensembleModel is  : ", accuracy_score(y_test, y_predictions))
print("Accuracy score using ensembleModel is  : ", confusion_matrix(y_test, y_predictions))
print("Accuracy score using ensembleModel is  : ", classification_report(y_test, y_predictions))
```
