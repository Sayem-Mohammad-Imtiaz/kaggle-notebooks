```python
# read imu.csv & convert to numpy array
import os
files = sorted(os.listdir('../input/acc2csv000853128/'))
print(files)
```

```python
data_path = '../input/acc2csv000853128/'
```

```python
import numpy as np

def read_acc(filename):
    f = open(data_path+filename,'r')
    line = f.readlines()
    #print(line)
    acc = np.fromstring(line[0], dtype=float, sep=',')
    print(len(acc))
    acc = acc.reshape(int(len(acc)/3),3)
    #print(acc)
    print(acc.shape)
    print(len(acc[:,0]))
    return acc
```

```python
import matplotlib.pyplot as plt

def plot_acc(acc):
    x = np.linspace(0,len(acc[:,0]),len(acc[:,0]))
    print(len(x))
    plt.plot(x, acc[:,0])
    plt.plot(x, acc[:,1])
    plt.plot(x, acc[:,2])
    plt.legend(['accX', 'accY', 'accZ'], loc='upper right')
    plt.show()
```

```python
for file in files:
    acc = read_acc(file)
    plot_acc(acc)
```
