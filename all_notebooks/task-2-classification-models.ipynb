# Splitting Data 

```python
target = heart['target']
variables = heart.drop('target',axis=1)
variables.head()

x_train,x_test,y_train,y_test = train_test_split(variables, target, test_size=0.2, random_state=42)
```

```python
x_train=np.array(x_train)
y_train=np.array(y_train)
x_test=np.array(x_test)
y_test=np.array(y_test)


print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

# ***Different Classification Models***

# Decision Tree

```python
clf = DecisionTreeClassifier(max_depth=15, random_state=12)

clf.fit(x_train, y_train)

clf_score_train = clf.score(x_train, y_train) #Accuracy
clf_score_test =clf.score(x_test, y_test)


print('train accuracy: {}%'.format(clf_score_train*100)) 
print('test accuracy: {}%'.format(clf_score_test*100))
```

```python
#from sklearn import tree
#tree.plot_tree(clf)
```

# Random Forest

```python
rf = RandomForestClassifier(n_estimators=7, max_depth=7)
rf.fit(x_train, y_train)

clf_score_train = rf.score(x_train, y_train) #Accuracy
clf_score_test =rf.score(x_test, y_test)

print('train accuracy: {}%'.format(clf_score_train*100)) 
print('test accuracy: {}%'.format(clf_score_test*100))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter = 10000)
lr.fit(x_train,y_train)

y_pred = lr.predict(x_test)

score_train = lr.score(x_train, y_train) #Coefficient of determination
score_test =lr.score(x_test, y_test)


print('train score: {}'.format(score_train)) 
print('test score: {}'.format(score_test))
```

# SVC

```python
from sklearn.svm import SVC
scaler=MinMaxScaler()

Sc_X_train=scaler.fit_transform(x_train)
Sc_X_test=scaler.fit_transform(x_test)

clf=SVC(kernel='rbf')
clf.fit(Sc_X_train, y_train)

clf_score_train = clf.score(Sc_X_train, y_train) #accuracy
clf_score_test =clf.score(Sc_X_test, y_test)

print('train accuracy: {}%'.format(clf_score_train*100)) 
print('test accuracy: {}%'.format(clf_score_test*100))
```
