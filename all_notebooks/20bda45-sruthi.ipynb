### Linear Regression is not  a good model

```python
from sklearn.tree import DecisionTreeRegressor
from matplotlib import pyplot
model=DecisionTreeRegressor(random_state=1234456)

# Fit the model
model.fit(x_train,y_train)

# Get importance
importance= model.feature_importances_

# Summarize feature importance
for i,v in enumerate(importance):
    print("Feature: %0d, Score : %0.5f" %(i,v))
    
# Plot feature importance
pyplot.bar([x for x in range(len(importance))],importance)
pyplot.show()
```

```python
a=dt_2.columns.tolist()
a.remove('UltimateIncurredClaimCost')
print(a)
```

```python
dict(zip(a,glm.coef_))
```

```python
imp=dict(zip(a,importance))
```

```python
imp_dt1=pd.DataFrame(list(imp.items()),columns=["feature","importance"])
```

```python
imp_data=imp_dt1.sort_values(by = 'importance')
imp_data
```

```python
col=dt_2.columns
col
```

```python
def feature_train_test_score(dataset,list_columns_drop):
    X_scale=pre.minmax_scale(dataset.drop(list_columns_drop,axis=1))
    Y=dt_2['UltimateIncurredClaimCost']
    x_train,x_test,y_train,y_test=ms.train_test_split(X_scale,Y,test_size=0.3,random_state=1234456)
    glm=lm.LinearRegression()
    glm.fit(x_train,y_train)
    diction={}
    diction['Test score']=glm.score(x_test,y_test)
    diction['Train score']=glm.score(x_train,y_train)
    return(diction)
```

```python
# Function to find test and train scores by including 
#  features that have importance greater than a given threshold
def scores(dataset,a):
    unimportant_features=imp_data.feature[0:a+1].tolist()
    unimportant_features.append('UltimateIncurredClaimCost')
    features=(np.setdiff1d(col,unimportant_features)).tolist()
    n=len(features)
    scores=feature_train_test_score(dataset,unimportant_features)
    output={}
    output['Number of features']=n
    output['Features']=features
    output['scores']=scores
    return(output)
```
