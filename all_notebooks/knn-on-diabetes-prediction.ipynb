```python
# Following Notebook includes K-Nearest Neighbors algorithm which predicts near to accurate values more than any other algorithms.
# Here 2 types of scores are calculated i.e. first score is of variable defined to train and test the data and other score depends upon test values and predicted values
# starting with columns from pregnancies to age:

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

df=pd.read_csv("../input/diabetes-prediction/Diabetespred.csv")
print(df)

X=df.loc[:,'Pregnancies':'Age']
Y=df['Outcome']

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=4)

neighbors=KNeighborsClassifier(n_neighbors=4)
neighbors.fit(X_train,Y_train)
print("The accuracy score is {}".format(neighbors.score(X_train,Y_train)))
pred=neighbors.predict(X_test)
print(pred)
print(Y_test)
print(accuracy_score(Y_test,pred))
```
