```python
import pandas as pd 
import numpy as np 
import datetime
import seaborn as sns 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
netflix = pd.read_csv('../input/netflix-original-films-imdb-scores/NetflixOriginals.csv')
```

```python
netflix.info()
```

```python
#checking for nulls
netflix.isnull().sum()
```

```python
netflix.head()
```

```python
#converting premier to dates
netflix['Premiere'] = pd.to_datetime(netflix['Premiere'])
```

```python
#creating month, year, and day of week cols for movies
netflix['month'] = netflix['Premiere'].apply(lambda time: time.month)
netflix['day of week'] = netflix['Premiere'].apply(lambda time: time.dayofweek)
netflix['year'] = netflix['Premiere'].apply(lambda time: time.year)
```

```python
netflix.head(1)
```

```python
#mapping months to data
Mmap = {1: 'Jan',2:'Feb',3:'Mar',4:'April',5:'May',6:'Jun',7:'July',8:'Aug',9:'Sept',10:'Oct',11:'Nov',12:'Dec'}
netflix['month'] = netflix['month'].map(Mmap)
```

```python
#mapping days of the week to data 
dmap = {0: 'Mon',1:'Tues',2:'Wed',3:'Thurs',4:'Fri',5:'Sat',6:'Sun'}
netflix['day of week'] = netflix['day of week'].map(dmap)
```

```python
netflix.head(1)
```

```python
netflix.isnull().sum()
```

# EDA

```python
sns.pairplot(netflix)
```

```python
sns.heatmap(netflix.corr(),annot=True) #no strong correlations 
```

```python
plt.figure(figsize=(10,4))
sns.lineplot(x='month',y='IMDB Score',data=netflix,ci=None)
```

```python
netflix_year = netflix.query("month == 'Jun'")
sns.lineplot(data=netflix_year,x='year',y='IMDB Score',ci = None)
#spike in 2015 then dropoff
```

```python
bar_order = ['Mon','Tues','Wed','Thurs','Fri','Sat','Sun']
sns.barplot(x='day of week',y= 'IMDB Score',data=netflix,palette='Set1',order=bar_order) 
#Movies reviews on wednesday have highest IMDB score with smallest standard error
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='year',data=netflix,palette='Dark2') #2015 had second lowest amount of movies released 
```

```python
sns.scatterplot(x='Runtime',y='IMDB Score',data=netflix) #majority of scores come from movies with about 100 mins of runtime
```
