# The strength of currencies
I'm comparing the gold purchaising power for the different currencies, to see how their relative power evolved in time.

```python
df = pd.read_csv("/kaggle/input/monthly-gold-prices/1979-2021.csv")
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')

# When data is missing use the first reported data
df = df.fillna(method='bfill')

# Scale to have the most recent values all equal (current status quo)
df = df.div(df.iloc[-1]) 
df = df.apply(lambda x: 1/x)

# Normalize to one each quarter
df = df.div(df.sum(axis=1), axis=0)

# For visualizing better the y axis, show the year only at the first month
df['YEAR'] = df.index.year
df.loc[df.duplicated(['YEAR']), 'YEAR'] = ""

df.tail(10000).plot(kind='barh', x='YEAR',stacked=True, figsize=[15, 20], width=1) # tail() for testing purposes
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.xlim(0,1)
plt.show()
```

## Not a nice graph for turkish people, who see their purchaising power gettin smashed every year...

```python
# Now plotting it with oldest data normalized
df = pd.read_csv("/kaggle/input/monthly-gold-prices/1979-2021.csv")
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')

# When data is missing use the first reported data
df = df.fillna(method='bfill')

# Scale to have the oldest values all equal (1979's status quo)
df = df.div(df.iloc[0]) # >>> ONLY DIFFERENCE TO PREVIOUS CELL
df = df.apply(lambda x: 1/x)

# Normalize to one each quarter
df = df.div(df.sum(axis=1), axis=0)

# For visualizing better the y axis, show the year only at the first month
df['YEAR'] = df.index.year
df.loc[df.duplicated(['YEAR']), 'YEAR'] = ""

df.tail(10000).plot(kind='barh', x='YEAR',stacked=True, figsize=[15, 20], width=1) # tail() for testing purposes
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.xlim(0,1)
plt.show()
```

## This second graph helps better to visualize the evolution, without the Turkish lira squeezing all the least recent values

#### Note how the currencies which became stronger are the Japanese and Swiss ones, which are known to be the most deflationary economies in the last decades.
