## Bar Plot

```python
ppl['Geographic Area'] == 'AL'  # it returns True or False for condtition

ppl[ppl['Geographic Area']=='AL']  # it returns dataframe objects which has True condition

x = ppl[ppl['Geographic Area']=='AL']

x 

print('sum of Area : ', sum(x.poverty_rate))

print('Count of City for Area : ', len(x))

print('Average Porverty for Area : ', sum(x.poverty_rate)/len(x))

x.poverty_rate.mean()
```

```python
#Powerty rate each state

area_list = list(ppl['Geographic Area'].unique()) 
area_poverty_ratio = []
# type(area_poverty_ratio)  --> List

for i in area_list:
    x = ppl[ppl['Geographic Area']==i]
    area_poverty_rate = sum(x.poverty_rate)/len(x) # OR area_poverty_rate =x.poverty_rate.mean()
    area_poverty_ratio.append(area_poverty_rate)

df = pd.DataFrame({'area_list': area_list,'area_poverty_ratio':area_poverty_ratio}) #not sorted by area_poverty_ratio
new_index = df['area_poverty_ratio'].sort_values(ascending=False).index.values #  we sort the index values by area_poverty_ratio
sorted_data = df.reindex(new_index)


# visualization
plt.figure(figsize=(20,6))
sns.barplot(x=sorted_data['area_list'], y=sorted_data['area_poverty_ratio'])
plt.xticks(rotation= 45)
plt.xlabel('States')
plt.ylabel('Poverty Rate')
plt.title('Poverty Rate Given States')
plt.show()
```

# High school graduation rate of the population that is older than 25 in states


```python
hs = pd.read_csv('/kaggle/input/fatal-police-shootings-in-the-us/PercentOver25CompletedHighSchool.csv',encoding ="windows-1252" )
```

```python
hs.head()
```

```python
hs.percent_completed_hs.value_counts()  
```

```python
hs.percent_completed_hs.replace(["-"],0.0, inplace=True)  #we filling missing data 
```

```python
hs.percent_completed_hs.value_counts()  
```

```python
hs.info()
```

```python
hs.percent_completed_hs = hs.percent_completed_hs.astype(float)
```

```python
hs.info()
```

```python
hs.head()
```

```python
area_list = list(hs["Geographic Area"].unique())
area_percentage_list = []

for i in area_list:
    x = hs[hs["Geographic Area"] == i]
    area_percentage = x.percent_completed_hs.mean()
    area_percentage_list.append(area_percentage)
    
#sorting
df = pd.DataFrame({"area_list":area_list,"area_percentage":area_percentage_list})
new_index = df.area_percentage.sort_values(ascending = True).index.values
sorted_data_2 = df.reindex(new_index)

# visualization
plt.figure(figsize=(18,7))
sns.barplot(x=sorted_data_2['area_list'], y=sorted_data_2['area_percentage'])
plt.xticks(rotation= 45)
plt.xlabel('Areas')
plt.ylabel('High School Graduate Rate')
plt.title("Percentage of Given State's Population Above 25 that Has Graduated High School")
plt.show()
```

# Percentage of state's population according to races that are black,white,native american, asian and hispanic


## Horizontal Bar Plot

```python
share_race_city = pd.read_csv('/kaggle/input/fatal-police-shootings-in-the-us/ShareRaceByCity.csv', encoding="windows-1252")
```

```python
share_race_city.head()
```

```python
share_race_city.info()
```

```python
# Percentage of state's population according to races that are black,white,native american, asian and hispanic

#data correction 
share_race_city.replace(['-'],0.0,inplace = True)
share_race_city.replace(['(X)'],0.0,inplace = True)

# change data type string to float
share_race_city.loc[:,['share_white','share_black','share_native_american','share_asian','share_hispanic']] = share_race_city.loc[:,['share_white','share_black','share_native_american','share_asian','share_hispanic']].astype(float)

#data preparing for visualization
area_list = list(share_race_city['Geographic area'].unique())
white_rate = []
black_rate = []
native_american_rate = []
asian_rate = []
hispanic_rate = []
for i in area_list:
    x = share_race_city[share_race_city['Geographic area']==i]
    white_rate.append(sum(x.share_white)/len(x))
    black_rate.append(sum(x.share_black) / len(x))
    native_american_rate.append(sum(x.share_native_american) / len(x))
    asian_rate.append(sum(x.share_asian) / len(x))
    hispanic_rate.append(sum(x.share_hispanic) / len(x))

# visualization
f,ax = plt.subplots(figsize = (15,15))
sns.barplot(x=white_rate,y=area_list,color='green',alpha = 0.5,label='White' )
sns.barplot(x=black_rate,y=area_list,color='blue',alpha = 0.7,label='African American')
sns.barplot(x=native_american_rate,y=area_list,color='cyan',alpha = 0.6,label='Native American')
sns.barplot(x=asian_rate,y=area_list,color='yellow',alpha = 0.6,label='Asian')
sns.barplot(x=hispanic_rate,y=area_list,color='red',alpha = 0.6,label='Hispanic')

ax.legend(loc='upper right',frameon = True)     # legend location and frame
ax.set(xlabel='Percentage of Races', ylabel='States',title = "Percentage of State's Population According to Races ")
plt.show()
```

## Point Plot

```python
# high school graduation rate vs Poverty rate of each state

#simple normalization
sorted_data['area_poverty_ratio'] = sorted_data['area_poverty_ratio']/max( sorted_data['area_poverty_ratio'])
sorted_data_2['area_percentage'] = sorted_data_2['area_percentage']/max( sorted_data_2['area_percentage'])

#concating data (even if states are different order, concat metod uses index and the data merge correctly)
df = pd.concat([sorted_data,sorted_data_2['area_percentage']],axis=1) #axis=1 --> vertical merge
df.sort_values('area_poverty_ratio',inplace=True)

# visualize
f,ax1 = plt.subplots(figsize =(20,7))
sns.pointplot(x='area_list',y='area_poverty_ratio', data=df, color='green',alpha=0.5)
sns.pointplot(x='area_list',y='area_percentage', data=df, color='red',alpha=0.5)
plt.text(40,0.60,'high school graduate ratio',color='red',fontsize = 17,style = 'italic')
plt.text(40,0.52,'poverty ratio',color='green',fontsize = 17,style = 'italic')
plt.xlabel('States',fontsize = 15,color='blue')
plt.ylabel('Values',fontsize = 15,color='blue')
plt.title('High School Graduate  VS  Poverty Rate',fontsize = 20,color='blue')
plt.grid()
```

## Joint Plot

```python
# Visualization of high school graduation rate vs Poverty rate of each state with different style of seaborn code
# joint kernel density
# pearsonr= if it is 1, there is positive correlation and if it is, -1 there is negative correlation.
# If it is zero, there is no correlation between variables
# Show the joint distribution using kernel density estimation 
g = sns.jointplot(df.area_poverty_ratio, df.area_percentage, kind="kde", size=7)
plt.savefig('graph.png')
plt.show()
```

```python
# you can change parameters of joint plot
# kind : { “scatter” | “reg” | “resid” | “kde” | “hex” }
# Different usage of parameters but same plot with previous one
g = sns.jointplot("area_poverty_ratio", "area_percentage", data=df,size=5, ratio=3, color="r")
```
