# Playing with line plots in plotly

```python
import numpy as np 
import pandas as pd 
import plotly.offline as pyo
import plotly.graph_objs as go

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv('/kaggle/input/uk-fuel-price-weekly-statistics20032020/fuel price.csv')
data['Date'] = pd.to_datetime(data['Date'])
data.head()
```

```python
data.columns
```

```python
ulsd_price = go.Scatter(
                    x=data['Date'],
                    y=data['Pump price in pence/litre (ULSD)'],
                    mode='markers',
                    name='ulsd_price')


ulsd_duty = go.Scatter(
                    x=data['Date'],
                    y=data['Duty rate in pence/litre (ULSD)'],
                    mode='markers',
                    name='ulsd_duty')

data = [ulsd_price, ulsd_duty]

layout = go.Layout(title='Pump price in pence/litre (ULSD)',
                  xaxis=dict(title='Year'),
                  yaxis=dict(title='Price'))

fig = go.Figure(data=data, layout=layout)

pyo.iplot(fig)
```
