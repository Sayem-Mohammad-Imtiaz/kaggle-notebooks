```python
dataset = pd.read_csv('../input/archive.csv',low_memory=False,header=0)
```

```python
# Create a DateTime from the year and month
# make this our new index
dataset['DateTime'] = pd.to_datetime(dataset.Year.map(str) + '-' + dataset.Month.map(str))
dataset.set_index(['DateTime'])
```

```python
Annual = dataset.groupby(['DateTime']).mean()
```

```python
fig, ax = plt.subplots(1, 1)
pt_CO2fit = Annual['Carbon Dioxide (ppm)'].plot(ax=ax)
pt_CO2 = Annual['Seasonally Adjusted CO2 (ppm)'].plot(ax=ax)
```
