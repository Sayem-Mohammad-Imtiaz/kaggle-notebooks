# State-wise Vaccinations in India

#### Python version - https://github.com/VidyasagarMSC/vaccine-visualizer

```python
###################################################
# Pandas profiling for Cowin data
# Altair for Visualization
# ipywidgets for dropdown menu
##################################################
from pandas_profiling import ProfileReport
import altair as alt
alt.renderers.enable('kaggle')
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
##################################################
# State-wise vaccine doses data 
##################################################
df = pd.read_csv("/kaggle/input/covid19indiaorg-vaccinations-india-data/vaccine_doses_statewise.csv")
df.info()
df.head()
```

```python
df.columns.name = "date"

# Convert the date from dd/mm/yyyy format to the format that works with Altair for Visualization
new_cols = dict(zip(df.columns[1:], [pd.to_datetime(x,format="%d/%m/%Y") for x in df.columns[1:]]))
df_formatted = df.rename(columns= new_cols, inplace=False)
df_formatted.set_index("State", inplace=True)
df_formatted.head()

def f(state):
    data = df_formatted.loc[[state]]
    data = data.T.reset_index()
    data = pd.melt(data, id_vars=["date"]).rename(columns = {"date": "date", "value": "Total Individuals Vaccinated"})
    chart = alt.Chart(data).mark_area().encode(
        x="date:T",
        y=alt.Y("Total Individuals Vaccinated:Q", stack=None),
        color="State:N",
        strokeDash='State',
        tooltip = [alt.Tooltip('State:N'),
                   alt.Tooltip('date:T'),
                   alt.Tooltip('Total Individuals Vaccinated:Q')
                    ]
            ).interactive()
    return chart

interact(f, state=df_formatted.index)
```
