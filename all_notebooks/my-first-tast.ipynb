```python
#Reading the csv file using pandas
df = pd.read_csv('/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv')
#Assigning variables
country = df['country']
vaccines = df['vaccines']
people_vaccinated = df['people_vaccinated']
people_vaccinated_percent = df['people_vaccinated_per_hundred']
```

```python
"""vaccines Used in each country"""
country_list = []
vaccines_list = []
for i,c in enumerate(country):
    if c not in country_list:
        country_list.append(c)
        vaccines_list.append(vaccines[i])
        

for i,c in enumerate(country_list):
    print(c.rjust(20),':',vaccines_list[i])
```

```python
"""What country vaccinated most people?"""
print("{} was vaccinated {:,} people.".format(country[people_vaccinated.idxmax()],int(people_vaccinated.max())))
```

```python
"""Which country vaccinated largest percentage of its population"""
print("{} vaccinated {:}% of its population.".format(country[people_vaccinated_percent.idxmax()],people_vaccinated_percent.max()))
```
