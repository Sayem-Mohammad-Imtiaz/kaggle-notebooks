# Plan for EDA report

In this EDA report I'm planning to give a brief statistical summary of the world happiness report for Nicaragua from 2006 - 2019

## Importing data

```python
# importing data
world_happiness_report = pd.read_csv("/kaggle/input/world-happiness-report-2021/world-happiness-report.csv")
nicaragua = world_happiness_report[world_happiness_report["Country name"]=="Nicaragua"]
```

## Feature changes from 2016 to 2019

```python
fig, axs = plt.subplots(3, 2, constrained_layout=True)
fig.set_figwidth(12)
fig.set_figheight(10)
axs[0, 0].plot(nicaragua["year"], nicaragua["Life Ladder"])
axs[0, 0].set_title("Life Ladder")
axs[0, 1].plot(nicaragua["year"], nicaragua["Log GDP per capita"])
axs[0, 1].set_title("Log GDP per capita")
axs[1, 0].plot(nicaragua["year"], nicaragua["Log GDP per capita"])
axs[1, 0].set_title("Log GDP per capita")
```
