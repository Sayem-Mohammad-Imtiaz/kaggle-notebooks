```python
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt
from tabulate import tabulate
import seaborn as sns
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
# pd.option_context('display.colheader_justify', 'right')
# pd.set_option('display.max_columns',None)
# pd.set_option('display.html.border',1)
cols = ['battery_power', 'blue',   'clock_speed', 'price_range']
# cols = []
df = pd.read_csv('/kaggle/input/mobile-price-classification/train.csv', engine='c')
# print(df.info())
# print(df['px_height'].describe())
# print(tabulate(df.head(10), headers=df.columns))
# print(tabulate(pd.DataFrame(df['price_range'].head(10))))
priceRange = {0: 'low cost', 1: 'medium cost',
              2: 'high cost', 3: 'very high cost'}
df['price_range'] = df['price_range'].map(priceRange)

# print(df.isnull().any())  # it has no null values
print(df.info())

y = pd.DataFrame(df['price_range'])
x = df.drop(['price_range'], axis='columns')
# print(type(y), '\n', type(x))
# print(y, '\n', x)
# print(df.corr())


scaler = StandardScaler()
scaledData = pd.DataFrame(scaler.fit_transform(x), columns=['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g', 'int_memory', 'm_dep', 'mobile_wt',
                                                            'n_cores', 'pc', 'px_height',  'px_width',  'ram', 'sc_h',  'sc_w', 'talk_time', 'three_g', 'touch_screen',  'wifi'])
# print(scaledData.head())
# scaledDrop = scaledData.drop(['n_cores', 'pc', 'px_height',  'px_width',  'ram',
#                               'sc_h',  'sc_w', 'talk_time', 'three_g', 'touch_screen',  'wifi'], axis='columns')

# # seaborn
# sns.pairplot(df,hue="price_range")
# # sns.heatmap(scaledDrop.corr())
# plt.show()

# matplotlib
# # plt.scatter(df['px_height'],df['price_range'])
# plt.bar(df['price_range'],df['px_height'])
# # plt.xlabel('ram')
# # plt.ylabel('price range')
# plt.show()

mod1 = LogisticRegression()
mod2 = RandomForestClassifier()
mod3 = tree.DecisionTreeClassifier()

# a = cross_val_score(mod1, scaledData, y, cv=10).mean()
# b = cross_val_score(mod2, scaledData, y, cv=10).mean()
# c = cross_val_score(mod3, scaledData, y, cv=10).mean()
# print(a, '\n', b, '\n', c)

mod1.fit(scaledData, y)
# print(mod1.score(scaledData, y))
# print(mod1)
testData = pd.read_csv('/kaggle/input/mobile-price-classification/test.csv')
testData = testData.drop(['id'], axis='columns')
# print(testData.head())
testScaleData = pd.DataFrame(scaler.fit_transform(testData), columns=['battery_power', 'blue', 'clock_speed', 'dual_sim', 'fc', 'four_g', 'int_memory', 'm_dep', 'mobile_wt',
                                                                      'n_cores', 'pc', 'px_height',  'px_width',  'ram', 'sc_h',  'sc_w', 'talk_time', 'three_g', 'touch_screen',  'wifi'])
# print(testScaleData.head())
predictedData = pd.DataFrame(mod1.predict(
    testScaleData), columns=['price_range'])
print(predictedData)
priceRange = {'low cost': 0, 'medium cost': 1,
              'high cost': 2, 'very high cost': 3}
predictedData['price_range'] = predictedData['price_range'].map(priceRange)
print(predictedData)
predictedData.to_csv('predicted mobile prices',index=False)
```
