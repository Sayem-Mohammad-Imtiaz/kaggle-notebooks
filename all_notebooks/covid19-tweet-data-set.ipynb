```python
import pandas as pd
#to create a data frame of the csv file
```

```python
#read the data 
tweets = pd.read_csv('../input/covid19-tweets/covid19_tweets.csv')
tweets.head(5)
```

```python
#checking for nan 
tweets.isnull().sum()
```

```python
#deleted rows which had null values and reset the index
tweets=tweets.dropna()
tweets.reset_index(drop=True)
```

```python
user=tweets.groupby(['user_name','user_location']).size().to_frame('count').reset_index()
#grouped user name and location to get the count of each group 
```

```python
topusers=user.sort_values(['count'],ascending=False)
#finding who did most tweets and where he is from 
top_user=topusers.head(5)
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(24,10))

# Plot the exponential function
sns.barplot(x =top_user['user_name'] , y =top_user['count'] )

plt.xlabel("Top 5 user name",fontsize=12)

plt.ylabel("Number of tweets per user",fontsize=12)


plt.title("Bar chart for top users",fontsize=15)
plt.show()
#plot of top 5 user name
```

```python
#displaying the countries from where tweets have come using visualizations


geolocation=tweets.groupby(['user_location']).size().to_frame('count').reset_index()
toplocation=geolocation.sort_values('count',ascending=False)
toplocation
```
