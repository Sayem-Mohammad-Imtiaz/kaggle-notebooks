# Data Cleansing
* There's one column with data separated by a ';'
* these should be 13 separate columns:
    * Get column names
    * separate the data in each row and create a list of values
    * create new dataframe with separate columns
   

```python
df.columns.values
```

```python
columns = (df.columns.values[0]).split(';')
```

```python
new_df_dict = {}

for index, row in df.iterrows():
    lst = str(row.values[0]).split(';')
    new_df_dict[index]=lst[1:]
```

```python
data = pd.DataFrame.from_dict(new_df_dict, orient='index')
data.columns = columns[1:]
```

```python
data.head()
```

```python
data.describe()
```

```python
data = data.astype(float)
```

```python
data['age'] = data['age'] // 365
data.head()
```

# Model

```python
X = data.drop('cardio', axis=1)
y = data['cardio']
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
```

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape = (11,) ))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X, y, epochs=10, batch_size=32, validation_split = .2)
```
