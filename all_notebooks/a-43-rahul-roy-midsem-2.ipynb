# Q3. Write a Python function that takes a list and returns a new list with the square of unique elements of the first list.

```python
def square_uniques():
    arr = list(map(int, input().split()))
    arr2 = list(set(arr))
    print("Input Array Uniques :", arr2)
    sq_arr= list(map(lambda x: x**2, arr2))
    return sq_arr

square_uniques()
```
