## Filling Missing Values

```python
df1['Embarked'].value_counts()
```

```python
import seaborn as sns
```

```python
nan_list=['Pclass','Sex','Age','Embarked','Cabin']
```

```python
import matplotlib.pyplot as plt
k=1
f, axes = plt.subplots(4,1, figsize=(10,25), sharey = True)
for i in range(0,4):
    sns.countplot(x=nan_list[k],data=df1,ax=axes[i])
    k=k+1
```

```python
df1['Embarked']=pd.get_dummies(df1['Embarked'],drop_first=True)
```

```python
df1['Embarked']
```

```python
df1['Embarked'].fillna(df1['Embarked'].mode()[0],inplace=True)
df1['Age'].fillna(df1['Age'].mode()[0],inplace=True)
```

```python
df1.isnull().sum()
```

```python
df1=df1.drop(columns=['Cabin','Ticket','PassengerId','Name'],axis=1)
```

```python
df1['Sex']=pd.get_dummies(df1['Sex'],drop_first=True)
```

```python
df1
```

```python
y=df1.iloc[:,0]
x=df1.iloc[:,1:]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf=RandomForestClassifier()
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```
