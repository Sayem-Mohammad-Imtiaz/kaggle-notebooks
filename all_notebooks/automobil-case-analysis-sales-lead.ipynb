```python
#Create a new variable by combining the manufacturer name and model.
df['man-model']=df['Manufacturer']+str('-')+df['Model']
df['man-model']
```

```python
#Create a new variable and the values conditionally.
df['new-v']=np.where(
       (df['DistanceToDealer']<10.00),'very-near',
np.where(
      ((df['DistanceToDealer']>=10.00) & (df['DistanceToDealer']<50.00)),'near',
    np.where(
      ((df['DistanceToDealer']>=50.00) & (df['DistanceToDealer']<100.00)),'far',
     np.where((df['DistanceToDealer']>=100.00),'very far', 'NA'))))
    
        
df['new-v'].value_counts()
```

```python
def con1(x):
    if x < 10.00:
        return "very near"
    elif((x >=10.00) and (x<50.00)):
        return "near"
    elif ((x >=50.00) and (x<100.00)):
        return "far"
    elif (x >=100.00):
        return "very far"
    else:
        return "NA"
    
```

```python
new_v3=[]
for i in range(len(df['DistanceToDealer'])):
   new_v3.append(con1(i))
df['new_v3']=new_v3
df['new_v3'].value_counts()

#print(con1(100.12))
#df['DistanceToDealer']
```

```python
#Which manufacturer has the highest market share.
df4=df['Manufacturer'].value_counts().reset_index(name='car_count')
df4['pct']=df4['car_count']/1499*100
df4
```

```python
import matplotlib.pyplot as plt

#Tasks = [300,500,700]
fig, ax1 = plt.subplots(figsize = (24,12)) 
my_labels = df4['index']
#my_colors = ['lightblue','lightsteelblue','silver']
#my_explode = (0, 0.1, 0)
plt.pie(df4['pct'], labels=my_labels, autopct='%1.1f%%', startangle=15, shadow = True)
plt.title('index')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
 
sizes = df4['car_count']

my_labels = df4['index']
labels = ["%s" % i for i in my_labels ]
fig1, ax1 = plt.subplots(figsize=(20, 10))
fig1.subplots_adjust(0.3, 0, 1, 1)
 
theme = plt.get_cmap('copper')
ax1.set_prop_cycle("color", [theme(1. * i / len(sizes))
                             for i in range(len(sizes))])
 
_, _ = ax1.pie(sizes, startangle=90, radius=1800)
 
ax1.axis('equal')
 
total = sum(sizes)
plt.legend(
    loc='upper left',
    labels=['%s, %1.1f%%' % (
        l, (float(s) / total) * 100)
            for l, s in zip(labels, sizes)],
    prop={'size': 11},
    bbox_to_anchor=(0.0, 1),
    bbox_transform=fig1.transFigure
)
 
plt.show()
```
