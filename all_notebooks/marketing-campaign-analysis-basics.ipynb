Can build a function to enable to calculate conversion rate across any sub-segment want on the fly.

* Function that takes a DataFrame and list of column names and outputs the conversion rate across the      column(s)
* Building a conversion function
* where users were converted, then group by the list of user inputted column_names 
* count the number of unique converted users.

```python
def conversion_rate(dataframe, column_names):
    # Total number of converted users
    column_conv = data[data['converted'] == True]\
                   .groupby(column_names)['user_id'].nunique()


    # Total number users
    column_total =  data.groupby(column_names)['user_id'].nunique()
    
    # Conversion rate 
    conversion_rate = column_conv/column_total
    
    # Fill missing values with 0
    conversion_rate = conversion_rate.fillna(0)
    return conversion_rate
```

```python
#Test and visualize conversion function
#conversion_rate() function to calculate the conversion rate in marketing by date_served and age_group
#store results in age_group_conv
```

```python
# Calculate conversion rate by age_group # by using above conversion function#

age_group_conv = conversion_rate(data, ['date_served',
                                             'age_group'])

# Unstack and create a DataFrame
age_group_df = pd.DataFrame(age_group_conv.unstack(level = 1))

# Visualize conversion by age_group
age_group_df.plot()
plt.title('Conversion rate by age group\n', size = 16)
plt.ylabel('Conversion rate', size = 14)
plt.xlabel('Age group', size = 14)
plt.show()
```

It seems to be age group 19-24 and 24-30 years has highest conversion rate than other age groups

```python
# Calculate conversion rate by language displayed # by using above conversion function#

lang_group_conv = conversion_rate(data, ['date_served',
                                             'language_displayed'])

# Unstack and create a DataFrame
lang_group_df = pd.DataFrame(lang_group_conv.unstack(level = 1))

# Visualize conversion by age_group
lang_group_df.plot()
plt.title('Conversion rate by lang displayed group\n', size = 16)
plt.ylabel('Conversion rate', size = 14)
plt.xlabel('language_displayed', size = 14)
plt.show()
```

A/B testing for marketing
when is the right time to determine key success metrics?The earlier define what success looks like,the better it is

Facebook campaign was actually run as an A/B test. Half the compaign messages sent out were generic upsells to  product .while the other half contained personalized messaging around the users’ usage of the site.


```python
#Check to ensure users were allocated equally to the test and control groups
```

```python
 # Subset the DataFrame
df_facebook = data[data['marketing_channel'] == 'Facebook']

# Group the facebook DataFrame by variant 
alloc = df_facebook.groupby(['variant'])['user_id'].nunique()

# Plot a bar chart of the test allocation
alloc.plot(kind='bar')
plt.title('Personalization test allocation')
plt.ylabel('# participants')
plt.show()
```

* Selecting conversion rate as our key metrics for this test.
* It is highly important that evaluate whether or not conversion was higher in the personalization treatment  compared with the control.
* Measuring the difference between the key metric in the control and the treatment is the most important part of evaluating the success of an A/B test.

```python
# Group data by user_id and variant
subscribers = df_facebook.groupby(['user_id', 
                             'variant'])['converted'].max()

subscribers_df = pd.DataFrame(subscribers.unstack(level=1)) 

# Drop missing values from the control column
control = subscribers_df['control'].dropna()

# Drop missing values from the personalization column
personalization = subscribers_df['personalization'].dropna()

print('Control conversion rate:', np.mean(control))
print('Personalization conversion rate:', np.mean(personalization))

#can see that personalization converted users are hiher rate than the control. 
```

* Creating a lift function.
* Lift can be calculated by calculating as follows 
* Formula: (Treatment conversion rate - Control conversion rate)/Control conversion rate

```python
def lift(a,b):
    # Calcuate the mean of a and b
    a_mean = np.mean(a)
    b_mean = np.mean(b)
    
    # Calculate the lift using a_mean and b_mean
    lift = (b_mean-a_mean)/a_mean
  
    return str(round(lift*100, 2)) + '%'
# Print lift() with control and personalization as inputs
print(lift(control, personalization))      # control and personalizaiton object created above

```

There's a large lift, but are results statistically significant that can be verfiy by using t test

* The personalization variant outperformed the control,to determine whether the result is statistically         significant. 
* Statistical significance is vital to understanding whether a test showed a positive result by chance or if   it is reflective of a true difference between the variants.
* This can help marketing team to make an informed choice about whether to roll out the feature or not.

```python
# To check difference between the control and personalization statistically significant?

from scipy.stats import ttest_ind

t = ttest_ind(control, personalization)
print(t) 
```

The resultant pvalue very small(e-18) ,So result is significant and personaltion variant has impact on conversion rate as compared to generic messages.
