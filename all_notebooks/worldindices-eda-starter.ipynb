# Plot data for each tickers

```python
for ticker in full_data['ticker'].unique():
    df = full_data.query('ticker == @ticker').copy()
    df.plot(x='date', y='close')
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('price')
    plt.title('{} ({})'.format(
        ticker_map.query('ticker == @ticker')['name'].values[0], ticker)
    )
    plt.show()
```
