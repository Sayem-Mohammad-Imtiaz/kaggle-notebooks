```python
data = list(df["products"].apply(lambda x:x.split(','))) #Through this cell products are seperated
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

Using and TransactionEncoder object, we can transform this dataset into an array format suitable for typical machine learning APIs. Via the fit method, the TransactionEncoder learns the unique labels in the dataset, and via the transform method, it transforms the input dataset (a Python list of lists) into a one-hot encoded NumPy boolean array:
mo


if you want a more info look the this page:[http://rasbt.github.io/mlxtend/user_guide/preprocessing/TransactionEncoder/](http://)

```python
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data,columns=te.columns_)
#all its products are converted into a column
#data type is boolean
```

```python
df
```

```python
te_data
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
df_new = apriori(df,min_support=0.2,use_colnames=True,verbose=1)
```

min_support = Assigns those that meet the condition 0.2 to the new variable

```python
df_new
```

```python
df_new.sort_values(by="support", ascending = False)
```

```python
df_new['length'] = df_new['itemsets'].apply( lambda x:len(x))
```

```python
df_new
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
association_rules(df_new, metric ="confidence", min_threshold=0.2)
#if confidence values>2
```

```python
association_rules(df_new, metric="support", min_threshold=0.2)
##if support values>2
```

```python
association_rules(df_new, metric ="lift", min_threshold=1.0)
#if lift values>2
```
