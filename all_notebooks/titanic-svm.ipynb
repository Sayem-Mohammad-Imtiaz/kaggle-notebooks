```python
titanic=pd.read_csv('../input/titanic-dataset-from-kaggle/train.csv')
```

```python
titanic.head(10)
```

```python
titanic["Fare"]=titanic["Fare"].values.astype(int)
```

```python
titanic["Fare"].head(5)
```

```python
titanic.corr()
```

```python
titanic["Sex"].replace({"male": "0", "female": "1"}, inplace=True)
```

```python
titanic['Cabin']=titanic["Cabin"].fillna(0)
```

```python
mean=np.mean(titanic['Age'])
titanic['Age']=titanic['Age'].fillna(mean)
titanic['Age']=titanic['Age'].values.astype(int)
```

```python
titanic["Embarked"].replace({"C": "1", "S": "2","Q":"3"}, inplace=True)
titanic["Embarked"]=titanic["Embarked"].fillna(0)
```

```python
titanic.head(3)
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model, metrics, model_selection
```

```python
from sklearn import svm
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(titanic[["Age","Pclass","Parch","Sex","Embarked","Fare"]],titanic.Survived,test_size=0.2)
```

```python
model = svm.SVC(kernel='linear') # Linear Kernel
```

```python
model.fit(X_train,Y_train)
```

```python
X_test
```

```python
predict=model.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(Y_test, predict))
```
