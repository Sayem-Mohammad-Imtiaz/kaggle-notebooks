```python
f, ax = plt.subplots(1,2,figsize = (18,8))
df_challenger[['blueFirstBlood','blueWins']].groupby(['blueFirstBlood'],as_index=True).mean().plot.bar(ax=ax[0])
ax[0].set_title('blueFirstBlood vs blueWins')
sns.countplot('blueFirstBlood',hue='blueWins',data = df_challenger,ax = ax[1])
ax[1].set_title('blueFirstBlood: BlueWins vs RedWins')
plt.show()
```

```python
df_challenger['Real_Game']=0
```

```python
fig, ax = plt.subplots(1,1,figsize=(9,5))
sns.kdeplot(df_challenger[df_challenger['blueFirstDragon']==1]['gameDuraton']/60,ax = ax)
sns.kdeplot(df_challenger[df_challenger['blueFirstDragon']==0]['gameDuraton']/60,ax = ax)
plt.legend(['blueFirstDragon = 1','redFirstDragon = 0'])
plt.show()
```
