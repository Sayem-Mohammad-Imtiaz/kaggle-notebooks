### Problem Statement

we are going to analyze the restaurant business data. We want to study the food habbits of zomato users, their preferances, ratings etc.



### Import Libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

Read the data and merge the two diferent datasets

```python
df = pd.read_csv('/kaggle/input/zomato-restaurants-data/zomato.csv',encoding="ISO-8859-1")
```

```python
df.head()
```

```python
country = pd.read_csv('/kaggle/input/country-code/country_code.csv')
country.head()
```

```python
data = pd.merge(df, country, on='Country Code')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

Number of Restaurants by Country on Zomato

```python
# No of restaurant that are on zomato in each country

countries = data['Country'].value_counts().index

no_of_restaurent = data['Country'].value_counts()

plt.figure(figsize=(15,10))
plt.bar(countries, no_of_restaurent)
plt.xlabel('Country', fontsize=20)
plt.ylabel('No of Restaurant', fontsize=20)
plt.title('No of Restaurant', fontsize=30, color='navy')
for idx, num in enumerate(no_of_restaurent):
    plt.text(idx-0.3, num + 30, s=str(num), color='red')
plt.xticks(rotation = 60)
plt.show()
```

Which city has the maximum number of Restaurents?

```python
data[data['Country'] !='india'].pivot_table(index=['Country','City'],
                                            values='Restaurant ID', aggfunc='count').sort_values('Restaurant ID', ascending=False)
```

```python
data_City = data[data['Country'] =='India']
Total_city =data_City['City'].value_counts()
Total_city.plot.bar(figsize=(14, 7),fontsize=14)
plt.title('Restaurants by City', fontsize=30, color='navy')                                             
plt.xlabel('City', fontsize=20)
plt.ylabel('No of Restaurants', fontsize=20)
plt.show()
```

What is the condition of online delivery in India ?

```python
data_online_delivery = data[data['Country'] == 'India']

data_online_delivery['Has Online delivery'].value_counts().plot.bar(figsize=(14, 7), 
                                             fontsize=14)
plt.xlabel('Country', fontsize=20)
plt.ylabel('No of Restaurants', fontsize=20)
plt.title('Restaurants available for online orders', fontsize=30, color='navy')
plt.show()
```

What are the top cuisines on Zomato?

```python
fig = plt.figure(figsize=(12,7))
sns.barplot(data['Cuisines'].value_counts().index[:15], data['Cuisines'].value_counts()[:15])
plt.xlabel('Cuisines', fontsize=20)
plt.ylabel('Number of Resturants', fontsize=20)
plt.title('Top 15 Cuisines on Zomato', fontsize=30, color='navy')
plt.xticks(rotation = 90)
plt.show()
```

```python
cuisines = ['North Indian', 'Fast Food', 'Chinese']
```

```python
data_small = data[data['Cuisines'].isin(cuisines)]
```

```python
sns.catplot(x='Cuisines', col='Country', col_wrap=3, kind='count', data=data_small.dropna(), sharey=False)
```

Average Ratings on Countries

```python
country_average = data.groupby('Country')['Aggregate rating'].mean().reset_index()

plt.figure(figsize=(25,10))
plt.xlabel('Country', fontsize=20)
plt.ylabel('Average Ratings', fontsize=20)
plt.title('Average Ratings on Countries', fontsize=30, color='navy')
plt.bar(country_average['Country'], country_average['Aggregate rating'])
```

Plot the ratings histogram for Indian restaurents

```python
plt.hist(data[data['Country'] == 'India']['Aggregate rating'],bins = 20)
plt.xlabel('Ratings',fontsize = 15)
plt.ylabel('No. of Ratings',fontsize= 15)
plt.show()
```

More than 2000 restaurents have been rated 0. Check the rating text as well as the votes of these restaurants

```python
data[data['Aggregate rating'] == 0]['Votes'].value_counts().plot(kind = 'bar',figsize =(15,10))
plt.xlabel('Ratings')
plt.ylabel('Number of people who have give a rating')
plt.show()
```

Word cloud for the most popular restaurants
