```python
def de_mean(x):
    x_bar = np.mean(x)
    return [x_i - x_bar for x_i in x]

def covariance(x, y):
    n = len(x)
    return np.dot(de_mean(x), de_mean(y)) / (n - 1)
```

```python
def correlation(x, y):
    std_x = np.std(x)
    std_y = np.std(y)
    if std_x > 0 and std_y > 0:
        return covariance(x, y) / (std_x * std_y)
    else:
        return 0
```

```python
pop = transformed.POP.values
gdp = transformed.GDPCAP.values
```

```python
plt.scatter(pop, gdp)
```

***The relationship between GDP CAP (per capita GDP) and POP (point of purchase) is not linear therefore, it is wise to employ Spearman's correlation, however we would employ both the Pearson's and Spearman's method.***

```python
correlation(pop, gdp)
```

```python
def SpearmanCorrelation(x, y):
    xranks = pd.Series(x).rank()
    yranks = pd.Series(y).rank()
    return correlation(xranks, yranks)
```

```python
SpearmanCorrelation(pop, gdp)
```
