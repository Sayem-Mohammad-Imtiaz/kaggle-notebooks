## Preprocessing of data

```python

df_bike['day'] = pd.to_datetime(df_bike['datetime']).dt.date
df_bike['hour'] = pd.to_datetime(df_bike['datetime']).dt.time
```

```python
df_bike.head()
```

```python
df_bike_per_day =df_bike.groupby('day').mean()
```

```python
df_bike_per_day=df_bike_per_day.reset_index()
df_bike_per_day.head()
```

```python
df_bike_per_day.isna().mean()*100
```

## Some Explortion in data for further use

let's see the distribution of 'count' column which presents the usage number of bikes

```python
#bin_values = np.arange(start=-50, stop=200, step=10)
#df_bike_per_day['count'].hist(bins=bin_values, figsize=[14,6])
plt.figure(figsize=(10,6))
plt.title("Distribution of average of bike usage per day", fontsize=15)
sns.distplot(df_bike_per_day['count'])
```

Let's create a new column in the dataframe which contains the range of bike usage per day

```python
df=df_bike_per_day
df['range_usage']= "unknown"
for i,n in enumerate(df['count']):
  n=round(n)
  if n<=100:
    df.loc[i,'range_usage']="[0-100["
  elif 100<n<=200:
    df.loc[i,'range_usage']="[100-200["
  elif 200<n<=300:
    df.loc[i,'range_usage']="[200-300["
  elif 300<n<=400:
    df.loc[i,'range_usage']="[300-400["
  elif n>400:
    df.loc[i,'range_usage']="[400-*["
```

we will now remove count column as  we have the range_usage column

```python
df.drop(['count','day'],axis=1,inplace=True)
df.head()
```

let's inspect correlation between columns of this new dataframe

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(), annot=True, annot_kws={'size':10}, cmap='Reds' )
```

## Train your model (KNN)

```python
x=df.loc[:, df.columns != 'range_usage']

y=df['range_usage']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=30)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
n_neighbors=20   
knn=KNeighborsClassifier(n_neighbors)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print('Accuracy=',accuracy_score(y_pred,y_test))
```
