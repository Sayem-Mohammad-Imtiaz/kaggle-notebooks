```python
# Top rated books (Revieved by at least 3000 people)
data_order = df.groupby(['Name', 'Author', 'Genre'], as_index=False)[['User Rating', 'Reviews']].mean()
data_order = data_order[data_order['Reviews']>3000]
data_order = data_order.sort_values('User Rating', ascending=False).head(30)
data_order
```

```python
# Weighted rating
m = min(df['Reviews'])
C = df['User Rating'].mean()
def weighted_rating(x, m=m, C=C):
    v = x['Reviews']
    R = x['User Rating']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
df['Weighted Rating'] = df.apply(weighted_rating, axis=1)
```

```python
top=df.groupby(['Name','Author','Genre'], as_index=False)[['User Rating', 'Reviews', 'Weighted Rating']].mean().sort_values(by='Weighted Rating', ascending=False).head(10)
top
```

```python
g=sns.barplot(x = top['Name'],
            y = top['Weighted Rating'],alpha=0.8)
plt.xticks(rotation=90)                 
plt.title("Top 10 books by Average Weighted Rating")
```

```python
g=sns.barplot(x = top['Author'],
            y = top['Weighted Rating'],alpha=0.8)
plt.xticks(rotation=90)                 
plt.title("Top 10 Authors by Average Weighted Rating")
```

```python
g=sns.barplot(x = top['Name'],
            y = top['Reviews'],alpha=0.8)
plt.xticks(rotation=90)                 
plt.title("Top 10 most reviewed books")
```
