# This section shows the the correlation in a heatmap for each of the csv files

There is a strong correlation between year and summer as I would expect that because of he severe winter battles are fought in the summers.



```python

correlation = got_battles.corr()
mp.figure(figsize=(10,10))
sns.heatmap(correlation, vmax=1, square=True,annot=True,cmap='cubehelix')

mp.title('Battle Analysis')
```

```python
# Character Death Analysis

From the data there seems to be higher correlation bewteen book of death and DwD.
```
