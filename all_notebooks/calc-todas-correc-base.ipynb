```python
def Find_Correc (CORREL, ACAO, ACAO_COR, BESTCORRE_df, CORRMAX, CORRMIN):
    TEMP = CORREL[ACAO].loc[((CORREL[ACAO] > CORRMAX) | (CORREL[ACAO] < CORRMIN)) & (CORREL[ACAO] != 1.0)]
    TEMP_df = pd.DataFrame(TEMP)
    TEMP_df[ACAO_COR] = TEMP_df.index
    TEMP_df.reset_index(drop=True, inplace=True)
    BESTCORRE_df[ACAO] = TEMP_df[ACAO]
    BESTCORRE_df[ACAO_COR] = TEMP_df[ACAO_COR]
    return(BESTCORRE_df)
```

```python
BESTCORRE_df = Find_Correc (CORREL, 'PETR3_PT', 'PETR3_PTCOR', BESTCORRE_df, CORRMAX, CORRMIN )
display (BESTCORRE_df)
```

```python
def TransColunas (BESTCORRE_Cols_df, ACAO, ACAO_COR, BESTCORRE_df):

    TEMP_df = pd.DataFrame()
    TEMP_df['INDIC_COR'] = BESTCORRE_df[ACAO]
    TEMP_df['ACAO_COR'] = BESTCORRE_df[ACAO_COR]
    TEMP_df['ACAO'] = ACAO

    BESTCORRE_Cols_df = BESTCORRE_Cols_df.append(TEMP_df, ignore_index=True)
    BESTCORRE_Cols_df = BESTCORRE_Cols_df.fillna(0)
    BESTCORRE_Cols_df = BESTCORRE_Cols_df[BESTCORRE_Cols_df['ACAO_COR'] != 0]
    
    return (BESTCORRE_Cols_df)
```
