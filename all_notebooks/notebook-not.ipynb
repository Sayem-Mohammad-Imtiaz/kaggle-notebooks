```python
team = pd.read_csv("/kaggle/input/skillsay-ai-crickethon-dataset/matches.csv")
team.info()
```

Q1. Highest leading run to win per season.

```python
seasan = team.groupby('season')['win_by_runs'].max()
seasan
```

```python
seasan = team.groupby('season')['win_by_runs'].max()
x = list(dict(seasan.sort_values(ascending=False).head(20)))
y = list(seasan.sort_values(ascending=False).head(20))
plt.bar(x,y)
plt.show()
```

Q2. Teams who has elected to bat first and has won the match.

```python
elected_bat = team[(team['toss_decision']=='bat')
                   & (team['toss_winner']==team['winner'])]
elected_bat.sort_values(by=['season'])
```

Q3. Top five players being awarded the player of the match award most number of times.

```python
count_awards = team.groupby('player_of_match')['season'].count()
count_awards.sort_values(ascending=False).head(5)
```

```python
x = list(dict(count_awards.sort_values(ascending=False).head(5)))
y = list(count_awards.sort_values(ascending=False).head(5))
plt.bar(x,y)
plt.show()
```

Q4. Number of matches played at each venue.

```python
venue = team.groupby('venue')['venue'].count()
venue.sort_values(ascending=False)
```

Q5. Number of tosses won by a team.

```python
toss_winner = team.groupby('toss_winner')['toss_winner'].count()
toss_winner.sort_values(ascending=False)
```

Q6. Matches where dls method was applied.

```python
dl_applied = team[team['dl_applied']!=0]
dl_applied.sort_values(by=['season'])
```

Q7. Number of dls method applied per season.

```python
dl_per_season = team.groupby('season').agg({'dl_applied': np.sum})
dl_per_season.sort_values(by=['season'])
```

Q8. Top 3 upmires with highest number of matches.

```python
umpire = (team.groupby('umpire1')['season'].count() 
          + team.groupby('umpire2')['season'].count())
umpire.sort_values(ascending=False).head(3)
```

```python
x = list(dict(umpire.sort_values(ascending=False).head(3)))
y = list(umpire.sort_values(ascending=False).head(3))
plt.bar(x,y)
plt.show()
```

Q9. Maximum number of matches won by team per season.(60 shown) 

```python
matches_won = team.groupby(['season', 'winner'])['season'].count()
matches_won.sort_values().head(60)
```

Q10. Number of draws season-wise.

```python
draws = team[team['result']!='normal']
draws.sort_values(by=['season'])
```
