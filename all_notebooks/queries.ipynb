```python
import numpy as np
import pandas as pd
from datetime import *
```

```python
df = pd.read_csv('../input/forbes-top-billionaires-list-2018/forbes_2018.csv')
```

```python
df = df[52:]
df.reset_index(drop=True,inplace=True)
df.drop(['position','lastName','age','gender','wealthSource','industry','worthChange','realTimeWorth','realTimePosition','image'],axis=1,inplace=True)
```

```python
df.head()
```

```python
top_ten = {}
for i in df.country.unique():
    top_ten[i] = df.loc[df.country==i,'worth'][:10].sum()
```

```python
dfx = pd.DataFrame.from_dict(top_ten,orient='index',columns=['Total_Net_worth'])
dfx.index.rename('Country',inplace=True)
```

```python
retail = pd.read_csv('../input/retailtransactiondata/Retail_Data_Transactions.csv')
```

```python
retail.drop('customer_id',axis=1,inplace=True)
retail.index.rename('transaction_id',inplace=True)
```

```python
retail
```

```python
retail.trans_date = pd.to_datetime(retail['trans_date'])
```

```python
dff = retail.groupby(retail['trans_date']).sum()
```

```python
dff['tran_date'] = dff.index
dff.reset_index(drop=True,inplace=True)
dff.index.rename('transaction_id',inplace=True)
```

```python
trans_diff = {}
trans_diff['2011-05-16'] = 5755
for i in dff.tran_date:
    diff = dff.loc[dff.tran_date == i,'tran_amount'].sum() - dff.loc[dff.tran_date == i + timedelta(days=1),'tran_amount'].sum()
    trans_diff[(i + timedelta(days=1)).date()] = diff
```

```python
result = pd.DataFrame.from_dict(trans_diff,orient='index',columns=['changed_by'])
result.index.rename('date',inplace=True)
result = result.head(-1)
```
