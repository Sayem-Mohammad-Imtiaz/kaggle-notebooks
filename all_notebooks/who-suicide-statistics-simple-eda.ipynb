## 20 Country with Most Suicide Case All the Time

```python
# 20 Country with most suicide case all the time
top_country = df.groupby('country').suicides_no.sum().sort_values(ascending=False).head(20)
top_country
```

```python
# Plot top_country data
plt.figure(figsize=(20,5))
ax = sns.barplot(top_country.index, top_country.values)
plt.xticks(rotation= 70)
ax.set_title('20 Country with Most Suicide Case (1979 - 2016)', size = 15)
ax.set_xlabel('Country', size = 15)
ax.set_ylabel('Total Case', size = 15)
```

Russian Federation, USA and Japan are the country with most suicide case

```python
# Add new column of Suicide no/ population
df['Suicide per Population'] = df['suicides_no']/df['population']
df.head()
```

```python
# 20 Country with most suicide/ population
top_country_ratio = df.groupby('country')['Suicide per Population'].sum().sort_values(ascending=False).head(20)
top_country_ratio
```

```python
# plot top_country_ratio data
plt.figure(figsize=(15,5))
ax = sns.barplot(top_country_ratio.index, top_country_ratio.values)
plt.xticks(rotation=70)
ax.set_title('20 Country with Suicide Case per Pupulation Ratio', size = 15)
ax.set_xlabel('Country', size = 15)
ax.set_ylabel('Suicide per Population Ratio', size = 15)
plt.show()
```

```python
# Suicide case over the world throughout the years (1979 - 2016)
suicide_per_year = df.groupby('year').suicides_no.sum()
suicide_per_year
```

```python
# plot suicides per year data
plt.figure(figsize=(20,7))
ax = sns.barplot(suicide_per_year.index, suicide_per_year.values, palette=['lightblue'])
ax.set_title('Suicide Case Per Year')
ax.set_xlabel('Year')
ax.set_ylabel('Total Case')
plt.show()
```

```python
def plot_trend(dataframe, feature, title, country):
    plt.style.use('ggplot')
    plt.figure(figsize=(20,100))
    
    for i, country in enumerate(country):
        plt.subplot(30,4, i+1)
        data = dataframe[dataframe['country']==country]
        sns.lineplot(x=data['year'], y=data[feature], label=feature)
        plt.xlabel('')
        plt.tick_params(axis='x', which='both',top=False, bottom=False, labelbottom=False)
        plt.title(country)
    
    plt.suptitle(title, y = 1.005)
    plt.tight_layout()
    plt.show()
```

```python
country = list(df['country'].unique())
plot_trend(df,'suicides_no', 'Suicides Case Trend', country)
```
