```python
data = pd.read_csv("../input/voice.csv")
```

```python
data.info()
```

```python
data.label = [1 if each == "male" else 0 for each in data.label] #Erkek ise 1 Kad覺n ise 0
```

```python
y = data.label.values
x_data = data.drop(["label"],axis=1) #Erkek ve Kad覺n Columns drop edildi.
```

```python
x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data)).values #normalizasyon yap覺ld覺.
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42) # 42 state 0.82 accuracy
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
print("test accuracy {}".format(lr.score(x_test,y_test)))
```
