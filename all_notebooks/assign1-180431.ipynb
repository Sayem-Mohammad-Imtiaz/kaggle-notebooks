# Adding Columns of Reasons for Crime Against Women

```python
pd.set_option('display.max_columns',None)
```

```python
df.head()
```

```python
def add_columns(df):
    for x in range(1,12):
        s = "reason"
        s=s +' '+ str(x)
        df[s]=0
        s = "reason"
    return df
add_columns(df)
```

```python
df.shape
```

```python
df.iloc[:,-12].dtype
```

```python
df.iloc[:,-12]=df.iloc[:,-12].astype(str)
```

```python
df.iloc[:,-12].dtype
```

```python
col = df['type of crime against women'].astype(str)
pieces = col.str.split(",")
for i in range(len(col)): 
    for k in pieces[i]:
        k.strip()
        if int(k)==0 :
            continue
        df.loc[int(i),'reason {}'.format(int(k))] =1
df.head()
```

```python
reason_murder=df["reason for murder"].value_counts()
reason_murder.drop([0],inplace =True)
age_grp_murder=df.loc[:,"child victims":"female victims"].apply(lambda x: x.sum())
murder_victims=age_grp_murder.sum()
```

```python
reason_for_crime_ag_women=df.loc[:,"reason 1":"reason 11"].apply(lambda x : x.sum())
reason_for_crime_ag_women
total_victims_for_CrimeAgainstWomen=df["adult victims"].sum()
```

```python
kidnapped=df.loc[:,"kidnapped child victims":"kidnapped female victims"].apply(lambda x : x.sum())
kidnapped_victims=kidnapped.sum()
```

```python
city_grps = df.groupby("city")
murder_victims_kp = city_grps.get_group("Kanpur").loc[:,"child victims":"female victims"].apply(lambda x: x.sum()).sum()
murder_victims_ghazbd =city_grps.get_group("Ghaziabad").loc[:,"child victims":"female victims"].apply(lambda x: x.sum()).sum()
murder_victims_lukn=city_grps.get_group("Lucknow").loc[:,"child victims":"female victims"].apply(lambda x: x.sum()).sum()
kidnapped_victims_kp =city_grps.get_group("Kanpur").loc[:,"kidnapped child victims":"kidnapped female victims"].apply(lambda x: x.sum()).sum()
kidnapped_victims_ghazbd =city_grps.get_group("Ghaziabad").loc[:,"kidnapped child victims":"kidnapped female victims"].apply(lambda x: x.sum()).sum()
kidnapped_victims_lukn =city_grps.get_group("Lucknow").loc[:,"kidnapped child victims":"kidnapped female victims"].apply(lambda x: x.sum()).sum()
CrimeAgainstWomen_victims_kp =city_grps.get_group("Kanpur").loc[:,"adult victims"].sum()
CrimeAgainstWomen_victims_ghazbd =city_grps.get_group("Ghaziabad").loc[:,"adult victims"].sum()
CrimeAgainstWomen_victims_lukn =city_grps.get_group("Lucknow").loc[:,"adult victims"].sum()
murder_list = [murder_victims_kp,murder_victims_ghazbd,murder_victims_lukn]
kidnapping_list = [kidnapped_victims_kp,kidnapped_victims_ghazbd,kidnapped_victims_lukn]
CrimeAgainstWomen_list = [CrimeAgainstWomen_victims_kp,CrimeAgainstWomen_victims_ghazbd,CrimeAgainstWomen_victims_lukn]
```

# Visualization Part

```python
plt.figure(figsize=(9,4))
# plt.rcParams["figure.figsize"] = (7, 3)
plt.xlabel("Reason for Crime against Women")
plt.ylabel("No. of crimes")
plt.title("plot for analysis of reasons of Crime against Women ")
reason_for_crime_ag_women.plot.bar(width=.65)
```

```python
# reason for CrimeAgainstWomen are majorly reason 7, reason 8, reason 10 
# reason 7 = Rape only(Sec. 376 or 511 IPC)
# reason 8 = Assault on Women with Intent to Outrage her Modesty (Sec. 354 IPC)
# reason 10 = Protection of Children from Sexual Offences Act
```

```python
plt.figure(figsize=(7,4))
plt.xlabel("Types of Kidnapped Victims")
plt.ylabel("No. of Kidnappings")
plt.title("plot for analysis of kidnapping with age group")
kidnapped.plot.bar(width=0.15)
```

```python
# kidnapping of child victims is significantly higher than adults
```

```python
plt.figure(figsize=(8,4))
plt.xlabel("Reason for Murder")
plt.ylabel("No. of Murder")
plt.title("plot for analysis of murder reason")
reason_murder.plot.bar(width=0.5)
```

```python
# for a very large no of murders reason of murder is unknown 
# trend for reason for murder goes like this Unknown reasons>Love Affairs>Petty Quarrels>Personal Vendetta>Family Dispute>Money Disputes>Property Disputes>Casteism 
```

```python
plt.figure(figsize=(5,5))
age_grp_murder.plot.bar(width=.45)
```

```python
# no of male victims are significantly higher then female and child victims 
# murders of child victims is significantly lower than adults 
```

```python
plt.figure(figsize=(10,5))
city = ["kanpur","ghaziabad","lucknow"]
width = 0.15
x_indexes = np.arange(len(city))
plt.bar(x_indexes-width,murder_list,width = width ,color="black",label='no. of murders')
plt.bar(x_indexes,kidnapping_list,width = width ,color="green",label="no. of kidnapping")
plt.bar(x_indexes+width,CrimeAgainstWomen_list,width = width ,color="maroon",label="no. of CrimeAgainstWomen")
plt.xticks(ticks=x_indexes,labels=city)
plt.legend()
```

```python
# we can see the trend that In all cities no of murders and CrimeAgainstWomen are significantly higher then kidnapping
# and we talk about murders then, trend is like kanpur<ghaziabad<lucknow
# and about kidnapping the trend goes like  kanpur<lucknow<ghaziabad
# and finally trend for CrimeAgainstWomen goes like this ,kanpur<ghaziabad<lucknow
```
