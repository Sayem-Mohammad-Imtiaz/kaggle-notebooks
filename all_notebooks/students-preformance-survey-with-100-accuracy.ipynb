Now after we finished our survey about students' performance and how the features in our dataset affected the students' scores, I'm going to build a simple machine learning model to predict the students' marks, depend on the features in this dataset.

before we going to build the model, I want to explain what I'll do in the next cells.

First I want to change the categorical values to numeric values because the model can't work with categorical values.

then I'm going to use different algorithms to train our model then we will see which algorithm got the best accuracy.

the algorithms that I'm going to use :

1. SVM ( support vector machine )
2. Random forest
3. linear regression


#Build the model

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
```

```python
data['gender'] = data['gender'].map({'female':0,
                                     'male':1})
data['race/ethnicity'] = data['race/ethnicity'].map({'group B': 0 ,
                                                     'group C' : 1 , 
                                                     'group A' : 2 , 
                                                     'group D' : 3 , 
                                                     'group E' : 4})
data['parental level of education'] = data['parental level of education'].map({'some high school': 0,
                                                                              'high school':1,
                                                                              'some college':2,
                                                                              "associate's degree":3,
                                                                              "bachelor's degree":4,
                                                                              "master's degree":5})
data['lunch'] = data['lunch'].map({'standard':0,
                                   'free/reduced':1})
data['test preparation course'] = data['test preparation course'].map({'none':0,
                                                                       'completed':1})
```

```python
features = ['gender','race/ethnicity','parental level of education','lunch',
            'test preparation course','math score','reading score','writing score']
features = data[list(features)]
label = data['total_score'].values
```

```python
x_train , x_test , y_train , y_test = train_test_split(features,label,test_size=0.30,random_state=114)
```

```python
print("X train shape is: ", x_train.shape)
print("Y train shape is: ", y_train.shape)
print("X test shape is: ", x_test.shape)
print("Y test shape is: ", y_test.shape)
```

```python
svr = SVR(kernel='rbf',degree=3,C=1.0)
```

```python
svr.fit(x_train,y_train)
```

```python
print("SVR score is: :",np.round(svr.score(x_test,y_test)*100,2))
```

```python
Random_Forest_Regressor = RandomForestRegressor()
```

```python
Random_Forest_Regressor.fit(x_train,y_train)
```

```python
print("RandomForestRegressor score is: :",np.round(Random_Forest_Regressor.score(x_test,y_test)*100,2))
```

```python
Linear_Regression = LinearRegression()
```

```python
Linear_Regression.fit(x_train,y_train)
```

```python
print("LinearRegression score is: :",np.round(Linear_Regression.score(x_test,y_test)*100,2))
```

Finally, after we use the three algorithms the accuracy was 98.8% for SVM, 99.76% for Random forest, and 100% for linear regression.

As we saw the best accuracy was for the linear **regression algorithm**.
