```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go
```

**<font color = DarkSlateBlue size = 5>Importing data</font>**

```python
df = pd.read_csv('../../kaggle/input/2021-olympics-medals-in-tokyo/Tokyo Medals 2021.csv')
```

**<font color = DarkSlateBlue size = 5>Data analysis</font>**

```python
df1 = df.sort_values(by = ['Total'],ascending = False).reset_index().head(10)

tr1 = go.Bar(x = df1['Country'],
             y = df1['Gold Medal'],
             name = 'Gold',
             marker = dict(color = 'Gold'))

tr2 = go.Bar(x = df1['Country'],
             y = df1['Silver Medal'],
             name = 'Silver',
             marker = dict(color = 'Silver'))

tr3 = go.Bar(x = df1['Country'],
             y = df1['Bronze Medal'],
             name = 'Bronze',
             marker = dict(color = 'Chocolate'))

data = [tr1,tr2,tr3]

layout = dict(title ='Top 10 Countries with the most Olympic Medals',
              title_x = 0.5,
              title_font = dict(size = 16, color = 'Maroon'),
              barmode = 'relative',
              xaxis = dict(title = 'Country'),
              yaxis = dict(title = 'Count')
                  
                  )
fig = go.Figure(data = data, layout = layout)
fig.show()
```

**<font color = DarkRed size = 4>Animated Map</font>**

```python
df1 = df.sort_values(by = ['Total'],ascending = False).reset_index()

fig = px.choropleth(df1,
                    locations = 'Country',
                    locationmode = 'country names',
                    color = 'Total',
                    color_continuous_scale = 'twilight',
                    hover_name = 'Country',
                    animation_frame = 'Country',
                    hover_data  = ['Rank By Total'])

fig.update_layout(title = 'World Countries with Tokyo Olympic Medals',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'MidnightBlue'),
                  geo = dict(showframe = False,
                             showcoastlines = False,
                             projection_type = 'equirectangular'
                            
                            )
                 
                 
                 )

fig.show()
```

```python
df1 = df.groupby("Country")['Gold Medal'].sum().sort_values(ascending = False).reset_index().head(10)

fig = px.bar(df1,x = 'Country', y = 'Gold Medal',color ='Gold Medal',color_continuous_scale = 'plasma',
             )
fig.update_layout(title = 'Top 10 Countries with the most Gold Medals',
                  title_x = 0.5,
                  title_font = dict(size = 16, color = 'DarkSlateBlue')
                 
                 )

fig.show()
```
