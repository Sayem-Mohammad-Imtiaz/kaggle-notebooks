# EDA

category of subject

```python
sns.set()
plt.figure(figsize=(15,5))
sns.countplot(x = "subject",hue= "is_paid", data= sample)
plt.title("categolize subject")
```

groupby free and paid. Distribution of subscribers

```python
# groupby free and paid

sample_paid = sample.groupby("is_paid").get_group(True)
sample_free = sample.groupby("is_paid").get_group(False)


#Distribution of subscribers
sns.set()
f, ax = plt.subplots(figsize=(15, 4))
sns.distplot(sample_free['num_subscribers'], color="b");
ax.xaxis.grid(False)
ax.set(ylabel="Frequency")
ax.set(xlabel="num_subscribers")
ax.set(title="subscibers distribution_free")
sns.despine(trim=True, left=True)
plt.show()

sns.set()
f, ax = plt.subplots(figsize=(15, 4))
sns.distplot(sample_paid['num_subscribers'], color="b");
ax.xaxis.grid(False)
ax.set(ylabel="Frequency")
ax.set(xlabel="num_subscribers")
ax.set(title="subscibers distribution_paid")
sns.despine(trim=True, left=True)
plt.show()
```

```python
#Distribution of costs
sns.set()
f, ax = plt.subplots(figsize=(15, 4))
sns.distplot(sample_paid['price'], color="b");
ax.xaxis.grid(False)
ax.set(ylabel="Frequency")
ax.set(xlabel="price")
ax.set(title="price distribution")
sns.despine(trim=True, left=True)
plt.show()
```

level

```python
sns.set()
plt.figure(figsize=(15,5))
sns.countplot(x = "level", data= sample_free)
plt.title("levels_free course", fontsize=18)

plt.figure(figsize=(15,5))
sns.countplot(x = "level", data= sample_paid)
plt.title("levels_paid course", fontsize=18)
```

# What are the best free courses by subject?


Judged the course with the most subscribers to be the best free courses by subject.

* Business Finance

"Bitcoin or How I Learned to Stop Worrying and Love Crypro" 

* Graphic Design

"Photoshop In-Depth: Master all of Photoshop's Tools Easily"

"Learn HTML5 Programming From Scratch"

```python
bestsub_f_bus = sample_free.groupby("subject").get_group("Business Finance").sort_values("num_subscribers", ascending= False).head(5)
bestsub_f_gra = sample_free.groupby("subject").get_group("Graphic Design").sort_values("num_subscribers", ascending= False).head(5)
bestsub_f_mus = sample_free.groupby("subject").get_group("Musical Instruments").sort_values("num_subscribers", ascending= False).head(5)
bestsub_f_web = sample_free.groupby("subject").get_group("Web Development").sort_values("num_subscribers", ascending= False).head(5)


sns.set()

plt.figure(figsize=(12,4))
plt.barh(bestsub_f_bus["course_title"],bestsub_f_bus["num_subscribers"] )
plt.title("The best free cources by subject(Business Finance)")

plt.figure(figsize=(12,4))
plt.barh(bestsub_f_gra["course_title"],bestsub_f_gra["num_subscribers"] )
plt.title("The best free cources by subject(Graphic Design)")

plt.figure(figsize=(12,4))
plt.barh(bestsub_f_mus["course_title"],bestsub_f_mus["num_subscribers"] )
plt.title("The best free cources by subject(Musical Instruments)")

plt.figure(figsize=(12,4))
plt.barh(bestsub_f_web["course_title"],bestsub_f_web["num_subscribers"] )
plt.title("The best free cources by subject(Web Development)")
```

# Q.How are courses related?

```python
sample["published_timestamp"] = pd.to_datetime(sample["published_timestamp"])
```

```python
sample["year"] = sample["published_timestamp"].dt.year
```

```python
sample_business = sample.groupby("subject").get_group("Business Finance").groupby("year").count()
sample_design = sample.groupby("subject").get_group("Graphic Design").groupby("year").count()
sample_musical = sample.groupby("subject").get_group("Musical Instruments").groupby("year").count()
sample_web = sample.groupby("subject").get_group("Web Development").groupby("year").count()

sns.set()
plt.figure(figsize=(14,6))


plt.plot(sample_business.index, sample_business["subject"], marker="o", label="Business Finance")
plt.plot(sample_design.index, sample_design["subject"], marker="o", label="Graphic Design")
plt.plot(sample_musical.index, sample_musical["subject"], marker="o", label="Musical Instruments")
plt.plot(sample_web.index, sample_web["subject"], marker="o", label="Web Development")
plt.xlabel("year", fontsize=18)
plt.ylabel("course_count", fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title("annual increase subject courses")
plt.tight_layout()
plt.legend()
```
