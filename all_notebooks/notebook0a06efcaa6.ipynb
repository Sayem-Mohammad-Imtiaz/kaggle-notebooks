```python
df = pd.read_csv("/kaggle/input/mba-admission/Admission.csv")
df.dtypes
```

```python
len(df)
```

```python
df_admit = df.loc[df['Decision'] == 'admit']
```

```python
df_notadmit = df.loc[df['Decision'] == 'notadmit']
```

```python
df_border = df.loc[df['Decision'] == 'border']
```

```python
x_admit_gpa = df_admit['GPA'].to_list()
y_admit_gmat = df_admit['GMAT'].to_list()
plt.scatter(x_admit_gpa, y_admit_gmat, c = 'green', label = 'admit')

x_notadmit_gpa = df_notadmit['GPA'].to_list()
y_notadmit_gmat = df_notadmit['GMAT'].to_list()
plt.scatter(x_notadmit_gpa, y_notadmit_gmat, c = 'red', label = 'notadmit')

x_border_gpa = df_border['GPA'].to_list()
y_border_gmat = df_border['GMAT'].to_list()
plt.scatter(x_border_gpa, y_border_gmat, c = 'blue', label = 'border')

plt.legend()

plt.xlabel('GPA')
plt.ylabel('GMAT')

plt.show()
```

The "border" set of points is concave.

```python
fig = plt.figure()
fig.suptitle('border - GPA', fontsize=14, fontweight='bold')
bx_border_gpa = plt.boxplot(df_border['GPA'])
```

The "border" row with the maximum GPA score is deleted.

```python
df = df.drop(df[(df.Decision == 'border') & (df.GPA == 3.5)].index)
len(df)
```

The "border" row with the minimum GMAT score is deleted.

```python
df = df.drop(df[(df.Decision == 'border') & (df.GMAT == 313)].index)
len(df)
```

```python
df.describe()
```

```python
# Average score in each category
df.groupby('Decision').mean()
```

```python
df.Decision.value_counts()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop('Decision', 1), df['Decision'], test_size = .2, random_state=10)
```

```python
X_train.shape[0]
```

```python
X_test.shape[0]
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
y_esti_tree = dtc.predict(X_test)
```

```python
# Display the first 5 results
y_esti_tree[1:5]
```

```python
len(y_esti_tree)
```

```python
dtc.score(X_test, y_test)
```

```python
my_submission = pd.DataFrame({'GPA': X_test.GPA, 'GMAT': X_test.GMAT, 'Decision': y_esti_tree})
my_submission.to_csv('submission.csv', index=False)
```
