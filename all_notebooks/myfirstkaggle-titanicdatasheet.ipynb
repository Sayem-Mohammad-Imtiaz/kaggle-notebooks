```python
def get_drop_data():
    # reading train data
    train = pd.read_csv('../input/titanic-subset/Titanic.csv')
    
    
    train.drop('Survived', 1, inplace=True)
    
    train.reset_index(inplace=True)
    train.drop('index', inplace=True, axis=1)
    
    return train
```

```python
drop_data=get_drop_data()
```
