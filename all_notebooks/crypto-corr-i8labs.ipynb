```python
#Import Cell
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from os import listdir
from os.path import isfile, join
```

```python
main = pd.DataFrame(pd.date_range("2020-01-01 23:59:59", periods = 365, freq='D'), columns = ["Date"])
```

```python
main_file_dir = "../input/cryptocurrencypricehistory"
arr = listdir(main_file_dir)
```

```python
for i in arr:
    file = pd.read_csv("../input/cryptocurrencypricehistory/"+str(i))
    df_temp = file.loc[(file['Date']>'2020-01-01') & (file['Date']<'2020-12-31'), ["Date", "Close"]]
    df_temp.rename(columns = {'Close':str(i)}, inplace = True)
    df_temp['Date'] = pd.to_datetime(df_temp['Date'], format = '%Y-%m-%d')
    main = main.merge(df_temp, on="Date", how="left")
```

```python
main.head()
```

```python
corr_matrix = main.corr()
corr_matrix
```

```python
sns.heatmap(corr_matrix);
plt.show()
```
