```python
#Importing necessary libaries

import pandas as pd
from keplergl import KeplerGl 
```

```python
df_mad = pd.read_csv('../input/madrid-smart-bins/papeleras_madrid.csv',delimiter=";")
#preprocessing dataset Madrid
# df['Origin Time'] = pd.to_datetime(df['Origin Time'])
df_mad['Place'] = df_mad.apply(lambda x:'Madrid', axis=1)
df_mad.head()
```

```python
df_mad = df_mad.rename(columns={'longitud':'Longitude','latitud':'Latitude'})
df_mad.head()
```

```python
map_1 = KeplerGl(height=600)
```

```python
map_1.add_data(data=df_mad, name='Papeleras de Madrid')
map_1
```

```python
# Saving the map and download for visualization
map_1.save_to_html(file_name='Papeleras_Madrid.html')
```
