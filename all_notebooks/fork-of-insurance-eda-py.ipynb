```python
import numpy as np 
import pandas as pd 
import os
import matplotlib.pyplot as pl
import seaborn as sns
import warnings
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
warnings.filterwarnings('ignore')
data = pd.read_csv('../input/insurance.csv')
```

## Tratamiento variables

```python
smoker = np.unique(data['smoker'])
smoker
```

```python
def map_smoker(smoker):
    if smoker == 'yes':
        return 1
    else:
        return 0
data['smoker'] = data['smoker'].apply(map_smoker)
data.head()
```

```python
sex = np.unique(data['sex'])
sex
```

```python
def map_sex(sex):
    if sex == 'female':
        return 1
    else:
        return 0
data['sex'] = data['sex'].apply(map_sex)
data.head(20)
```

```python
region = np.unique(data['region'])
region
```

```python
def map_region(region):
    if region == 'southwest':
        return 1
    elif region == 'northwest':
        return 2
    elif region == 'northeast':
        return 3
    elif region == 'southeast':
        return 4
    else:
        return 0
data ['region'] = data ['region'].apply(map_region)
data.head()
```

## Escalado de variables

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
dataScaled = mms.fit_transform(data[['age', 'children','smoker', 'bmi']])
data[['age', 'children','smoker','bmi']] = dataScaled
data[['age', 'children','smoker','bmi']].describe()
```

```python
data.corr()['charges'].sort_values()
```

## Regresión Lineal

```python
# Defino x e y
x = data.drop(['charges', 'sex', 'region'], axis = 1)
y = data.charges
```

```python
#Parto el dataset en train y test
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.30, 
                                                    train_size=0.70, random_state = 0)

#Entreno el modelo
lr = LinearRegression().fit(x_train,y_train)

y_train_pred = lr.predict(x_train)
y_test_pred = lr.predict(x_test)

#Resultado
print(lr.score(x_train,y_train))
print(lr.score(x_test,y_test))
```
