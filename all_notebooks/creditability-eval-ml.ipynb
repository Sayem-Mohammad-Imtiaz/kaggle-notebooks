```python
nRowsRead = 1000 # specify 'None' if want to read whole file
df = pd.read_csv('../input/cusersmarildownloadsgermancsv/german.csv', delimiter=';', encoding = "ISO-8859-2", nrows = nRowsRead)
df.dataframeName = 'german.csv'
nRow, nCol = df.shape
print(f'There are {nRow} rows and {nCol} columns')
df.head().style.set_properties(**{'background-color':'bisque',
                                     'color': 'purple'})
```

```python
!pip install evalml
```

```python
from evalml.automl import AutoMLSearch
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(columns=['Creditability'])
y = df['Creditability']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
automl = AutoMLSearch(X_train=X_train, y_train=y_train, problem_type='binary')
```

```python
automl.search()
```

```python
automl.rankings
```

```python
pipe = automl.best_pipeline
pipe.fit(X_train, y_train)
pred = pipe.predict(X_test)
```

```python
#https://evalml.featurelabs.com/en/v0.9.0/
automl.describe_pipeline(3)
```

#Select Best pipeline

```python
#https://evalml.featurelabs.com/en/v0.9.0/
pipeline = automl.best_pipeline
pipeline.score(X_test, y_test, ["f1"])
```
