#No Missing Values.

```python
#Code from Gabriel Preda
#plt.style.use('dark_background')
def plot_count(feature, title, df, size=1):
    f, ax = plt.subplots(1,1, figsize=(4*size,4))
    total = float(len(df))
    g = sns.countplot(df[feature], order = df[feature].value_counts().index[:20], palette= ('#32a852', '#a84e32', '#3242a8'))
    g.set_title("Number and percentage of {}".format(title))
    if(size > 2):
        plt.xticks(rotation=90, size=8)
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2.,
                height + 3,
                '{:1.2f}%'.format(100*height/total),
                ha="center") 
    plt.show()
```

#Marriage Style : 1-Arranged Marriage, 2-Flirt Marriage

```python
plot_count("Marriage Style", "Marriage Style", df,4)
```

#Status of Having a Child: 1-Yes, 2-No

```python
plot_count("Status of Having a Child", "Status of Having a Child", df,4)
```

#Education: 1-Primary, 2-High School, 3-Bachelor, 4-Msc or PhD; 

```python
plot_count("Education", "Education", df,4)
```
