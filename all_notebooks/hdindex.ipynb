```python
import numpy as np 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
from sklearn.model_selection import cross_validate
```

```python
hdi=pd.read_csv('/kaggle/input/human-development-index-hdi/HDI.csv')
hdi.head()
```

```python
hdi.info()
```

```python
hdi.info()
```

```python
hdi.describe().T
```

```python
hdi.isnull().sum()
```

```python
hdi = hdi.drop(['Country', 'Coverage'], axis=1)
```

**fill the missing values**

```python
for col in hdi.columns:
    hdi[col] = hdi[col].fillna(hdi[col].mean())
```

```python
hdi.isnull().sum()
```

# **drop unwanted data**

**select features and label**

```python
X = hdi.iloc[:,1:]
y =hdi.iloc[:,:1]
```

```python
print(f'The shape of X is {X.shape}\n\nThe shape of y is {y.shape}')
```

```python
X.columns
```

```python
y.columns
```

**split train test data**

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.4,random_state=1)
```

# model fitting

```python
model= LinearRegression().fit(X_train,y_train)
```

```python
cv_res=cross_validate(model,X_train,y_train,return_train_score=True)
cv_res['test_score'].mean()
```

```python
Pred = model.predict(X_test)
Pred
```

```python
mean_squared_error(y_test,Pred)
```

```python
mean_absolute_error(y_test,Pred)
```

```python
r2_score(y_test,Pred)
```
