First we collect our X data using NumPy, deleting a few rows that were filled with "NA". our first column is 1 = male, 0 = female. Our next three are 1 or 0 for different islands

```python
import numpy as np
import math as m
from sklearn.model_selection import train_test_split
import pandas as pd

df=pd.read_csv('../input/palmer-archipelago-antarctica-penguin-data/penguins_size.csv').values
n = df.shape[0]
X = np.zeros([n,8])
t_raw = np.zeros(n)

X[:,4:] = df[:,2:-1]
for i in range(n):
    if df[i,6] == 'FEMALE':
        X[i,0] = 1
    if df[i,1] == 'Torgersen':
        X[i,1] = 1
    if df[i,1] == 'Biscoe':
        X[i,2] = 1
    if df[i,1] == 'Dream':
        X[i,3] = 1
    if df[i,0] == 'Adelie':
        t_raw[i] = 0
    if df[i,0] == 'Chinstrap':
        t_raw[i] = 1
    if df[i,0] == 'Gentoo':
        t_raw[i] = 2
        
to_delete = []
for i in range(n):
    if m.isnan(X[i,4]):
        to_delete.append(i)
X = np.delete(X, to_delete, 0) #Find and delete problematic rows with 'NA'.
t = np.delete(t_raw, to_delete, 0)
print(X.shape)
print(t.shape)
```

```python
X_train, X_test, t_train, t_test = train_test_split(X, t, stratify=t) #test_size=0.3
#t_train = to_categorical(t_train, 3)
#t_test = to_categorical(t_test, 3) #convert to binary
```

```python
import tensorflow as tf
import keras
from keras.datasets import mnist
from keras.layers import Dense
from keras.models import Sequential
from keras.utils import to_categorical
from keras.optimizers import Adam
from keras.layers import Dense, Input
from keras.models import Model
```

```python
input_layer = Input(shape=(8))
x = Dense(64, activation = 'sigmoid')(input_layer)
output_layer = Dense(units = 3, activation = 'softmax')(x)

nn = Model(inputs = [input_layer], outputs = [output_layer])
nn.compile(optimizer='adam', loss = 'sparse_categorical_crossentropy', metrics =['accuracy'])
nn.summary()
nn_fit = nn.fit(X_train, t_train, batch_size=1, epochs = 20)
```

```python
model = Sequential()
model.add(Dense(units=64, activation='sigmoid', input_shape=(8,)))
model.add(Dense(units=3, activation='softmax'))
model.summary()
```

```python
model.compile(Adam(lr=0.01), loss='categorical_crossentropy', metrics=['accuracy']) #optimizer="Adam"
model.fit(X_train, t_train, batch_size=1, epochs=20)
loss, accuracy  = model.evaluate(X_test, t_test)
print(f'Test accuracy: {accuracy:.3}')
```
