Converting country and discipline abbreviations to full name

```python
countries_abbr = df.country.unique()
countries_abbr.sort()
countries_abbr
```

```python
full_discipline_names = {
    "DYN": "Dynamic With Fins",
    "DNF": "Dynamic Without Fins",
    "STA": "Static Apnea",
    "DYNB": "Dynamic With Bifins",
    "NLT": "No Limit",
    "VWT": "Variable Weight",
    "CWT": "Constant Weight",
    "CNF": "Constant Weight Without Fins",
    "FIM": "Free Immersion",
    "CWTB": "Constant Weight With Bifins"
}

full_countries_names = {
    "AT": "Austria", 
    "AU": "Australia",
    "BE": "Belgium",
    "BR": "Brazil",
    "CA": "Canada",
    "CN": "China",
    "CO": "Colombia",
    "CU": "	Cuba",
    "CZ": "Czech",
    "DE": "Germany",
    "DK": "Denmark",
    "FI": "Finland",
    "FR": "France",
    "GB": "United Kingdom",
    "GR": "Greece",
    "HR": "Croatia",
    "IT": "Italy",
    "JP": "Japan",
    "MC": "Monaco",
    "NL": "Netherlands",
    "NZ": "New Zealand",
    "PL": "Poland",
    "RU": "Russian Federation",
    "SE": "Sweden",
    "SI": "Slovenia",
    "US": "United States",
    "VE": "Venezuela" 
}

df["country_full"] = df["country"].apply(lambda x: full_countries_names[x])
```
