# Dropping irrelevant colums

```python
df_new = df_new.drop(['Daily Summary','Wind Bearing (degrees)','Summary','Precip Type','Temperature (C)','Wind Speed (km/h)','Visibility (km)','Pressure (millibars)'], axis = 1)
```

# Making our dataset ready to plot by setting Formatted Date as index and sampling monthly 'Apparent Temperature (C)' and 'Humidity' around mean to make visualisation compact

```python
df_new['Formatted Date'] = pd.to_datetime(df_new['Formatted Date'])
df_new = df_new.set_index('Formatted Date')
data_final = df_new[['Apparent Temperature (C)','Humidity']].resample('M', convention='start').mean()
data_final
```

```python
data_final.shape
```

# Plotting Variation of Apparent temparature and Humidity for the year 2006-2016

```python
plt.figure(figsize=(20,3));
plt.plot(data_final['Humidity'], label = 'Humidity', color = 'red',linestyle='dashed');
plt.plot(data_final['Apparent Temperature (C)'], label = 'Apparent temp.',color = 'green');
plt.title('Variation of Apparent temparature v/s Humidity', fontsize= 25);
plt.legend(loc = 0, fontsize = 12);
plt.xticks(fontsize = 15);
plt.yticks(fontsize = 13);
```

# Plotting variation of Apparent temparature and Humidity for each month over the year 2006-2016

```python
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
for i in range (1,13):
    data2 = data_final[data_final.index.month==i]
    plt.figure(figsize=(15,5))
    plt.plot(data2.loc['2006-01-01':'2016-01-01', 'Apparent Temperature (C)'], marker='o', linestyle='-',label='Apparent Temperature (C)',color = 'blue');
    plt.plot(data2.loc['2006-01-01':'2016-01-01', 'Humidity'], marker='o', linestyle='-',label='Humidity',color = 'red');
    plt.legend(loc = 0,fontsize = 15);
    plt.xlabel(months[i-1], fontsize = 15);
    plt.title('Humidity v/s Apparent Temperature',fontsize = 15)
    
```

# 2. Humidity remains nearly same across the time.
# 3. There is large variations in apparent temperature for some months whereas there is small variation in temperature for some            months during the provided time period.
# 4. Also the temperature increases as well as decreases for months of january, february, march, november and december. Hence they        show large variations.

> # # Hence Null Hypothesis Ho is rejected as the Apparent temperature and humidity compared monthly across 10 years of the data does not only increases due to Global warming. The humidity does not increases and the temperature increaes as well as decreases heavely for some months
