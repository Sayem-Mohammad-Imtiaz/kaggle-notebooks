# Olympics data analytics

In this notebook we will possibly see most of the visualization with respect to olympics

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

import warnings
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('/kaggle/input/olympic-games-2021-medals/Tokyo 2021 dataset.csv')
df
```

```python
df.columns = ['rank','team','gold','silver','bronze','total','total_rank','code']
```

```python
df.describe()
```

```python
df.info()
```

##  There are no null values in the dataset

## We will subset the top 10 and visualize it

```python
top_10 = df.head(10)
top_10
```

```python
medals = ['gold','silver','bronze']

top_10.plot(x ='team', y=medals,kind='bar', figsize = (10,10))
plt.show()
```

### From the above graph, the top ranked states are plotted. USA outnumbered the rest of the countries

```python
India = df[df['team']=='India']
India
```

## This is only to see India's position and ranking

##  We will see a scatter plot to see the total 

```python
px.scatter(df,x ='team', y='total')
```

Hover the chart to see more

##  A medals df is made with a condition to eliminate values that are 0

```python
medals_df = df.query('gold >0 and silver >0 and bronze>0')
```

```python
medals_df
```

```python
px.bar(medals_df,x ='team', y='total', color ='bronze')
```

```python
px.bar(medals_df,x ='team', y='total', color ='silver')
```

```python
px.bar(medals_df,x ='team', y='total', color ='gold')
```

```python
fig = px.sunburst(medals_df, path =['gold','team'])
fig.show()
```

```python
px.scatter(df, x="team", color="gold",hover_name="team", size="total")
```

```python
rank_df = df.query('rank>40')
rank_df
```

```python
rank_df.reset_index(drop= 'True', inplace = True)
```

```python
rank_df
```

```python
px.bar(rank_df,x ='team', y=medals)
```

```python
fig = px.scatter(rank_df, x="team", y=medals, size='total',hover_name="team")
fig.show()
```

```python
fig = px.box(rank_df, y="total")
fig.show()
```

```python
fig = px.box(medals_df, y="total")
fig.show()
```

```python
fig = px.scatter_3d(df, x='team', y='gold', z='silver', size_max=18, opacity=0.7)

# tight layout
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
```
