```python
pd.crosstab(data.Exited, data.NumOfProducts, margins=True, normalize='index').round(2).style.background_gradient(cmap='autumn_r')
```

<font color=blue>**OBSERVATION :** Customers who have subscribed to only one service are more prone to discontinue the service</font> 

```python
pd.crosstab(data.Exited, data.IsActiveMember, margins=True, normalize='index').round(2).style.background_gradient(cmap='autumn_r')
```

<font color=blue>**OBSERVATION :** Inactive customers are more prone to discontinue to the service</font> 

```python
pd.crosstab(data.Exited, data.Geography, margins=True, normalize='index').round(2).style.background_gradient(cmap='autumn_r')
```

<font color=blue>**OBSERVATION :** The customers in Spain are less prone to discontinue the service</font>
