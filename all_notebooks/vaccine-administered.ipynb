# Analysis of Total vaccine administered in India upto 02/06/2021

```python
#importing required libraries
import os,sys
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

```python
#Importing data file using pandas
filename='../input/total-vaccine-india-31052021/Total vaccine administered.csv'
data= pd.read_csv(filename)
data.head()
```

```python
data.columns
```

```python
#Checking the Null values in the dataset
data.isnull().sum()
```

```python
#Pairplotting the data
sns.pairplot(data)
```

Since pairplot is not much clear. we will try Inline plots, Scatterplots and other

```python
fig = plt.gcf()
fig.set_size_inches(12, 5)
sns.scatterplot(x='Total Individuals Vaccinated',y='Total Covaxin Administered',color='Red', data=data)
sns.scatterplot(x='Total Individuals Vaccinated',y='Total CoviShield Administered',color='Green',data=data)
sns.scatterplot(x='Total Individuals Vaccinated',y='Total Sputnik V Administered',color='Blue',data=data)
```

```python
fig = plt.gcf()
fig.set_size_inches(10,7)
sns.scatterplot(x='Total Individuals Vaccinated',y='Male(Individuals Vaccinated)',color='Blue',
data=data)
sns.scatterplot(x='Total Individuals Vaccinated',y='Female(Individuals Vaccinated)',color='Red',
data=data)
sns.scatterplot(x='Total Individuals Vaccinated',y='Transgender(Individuals Vaccinated)',color='Yellow',
data=data)
```

From this Scatterplot, It is very clear that Male individuals are getting more vaccinated.

```python
fig = plt.gcf()
fig.set_size_inches(12, 5)
sns.scatterplot(x='Total Individuals Vaccinated',y='Updated On',color='Red', data=data)
```

As we can see due to many distinct dates. We are going to create a dummy dataset by using last 10 dates.

```python
data1=data.tail(10)
data1.head()
```

```python
#Line plot
plt.figure(figsize=(10,8))
sns.lineplot(x='Updated On',y='Total Individuals Vaccinated',color='Red', data=data1)
```

We can clearly analyze here that individuals are vaccinated at steady rate.

```python
#Line plot
plt.figure(figsize=(15,5))
sns.lineplot(x='Total Individuals Vaccinated',y='First Dose Administered',color='Red', data=data)
sns.lineplot(x='Total Individuals Vaccinated',y='Second Dose Administered',color='Blue', data=data)
```

# From this analysis, It is very clear and understandable that Vaccination is at steady rate in India.
