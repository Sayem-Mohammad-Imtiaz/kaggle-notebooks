```python
f = open('../input/Iris.csv', 'r')
data = f.read()
print(data)
```

```python
split_data = data.split('\n')

no_header = split_data[1:len(split_data)-1]
split_no_header = [] 

for item in no_header:
    split_no_header.append(item.split(","))
    
flower_id_array = []
sepal_length_array = []
sepal_width_array = []
petal_length_array = []
petal_width_array = []
specie_array = []

    

print(split_no_header[0])

for item in split_no_header:
    flower_id_array.append(item[0])
    
for item in split_no_header:
    sepal_length_array.append(item[1])
    
for item in split_no_header:
    sepal_width_array.append(item[2])

for item in split_no_header:
    petal_length_array.append(item[3])
    
for item in split_no_header:
    petal_width_array.append(item[4])
    
for item in split_no_header:
    specie_array.append(item[5])
    
    
    
print(petal_length_array)

```
