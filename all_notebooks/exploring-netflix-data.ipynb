# Data Cleaning

```python
data['show_id'].unique()
```

Drop show_id column as it is not relevant to for analysis

```python
data.drop('show_id', axis = 1, inplace = True)
```

```python
data.head(3)
```

director and cast column has the most numbers of missing data. There are several ways we can go about this. We can fill in these values by getting information of directors and casts online. But that would be too Ambitious for this tiny project. For now, We'll just drop 'director' and cast column entirely. And make our data more easy to work with. # Baby Steps! 

```python
data.drop(columns = ['director','cast'], axis = 1, inplace = True)
```

```python
data.head()
```

For all the missing data in date_added column, We'll assume it was added on netflix on same year as release year & We'll take december as a month 

```python
data['date_added'] = data['date_added'].fillna(f"December 1, {str(data['release_year'].mode()[0])}")
```

```python
data['date_added'].isna().count()
```

We have total 7787 entries in our dataset and as the count of na is 7787 in 'date_added' column. We can conclude that we don't have any missing values in our data for 'date_added' colum

```python
#What are the most common ratings?

data['rating'].value_counts()
```

TV-MA is the most common type of rating in our data. We have 7 missing values for 'rating' column. So we'll fill those values with the most column rating type


```python
data['rating'].fillna('TV-MA', inplace = True)
```


Similarly, We have 507 entries where data for column 'country' is missing. Again, We can try to google where these movie/tv shows were produced and fill in these missing values or we can fill in these values by country producing the most movies/shows in our dataset

```python
#Top 5 countries producing this movies and tv shows
data['country'].value_counts().head()
```

```python
data['country'].fillna('United States', inplace = True)
```
