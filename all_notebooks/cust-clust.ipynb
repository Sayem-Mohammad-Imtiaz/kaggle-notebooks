```python
sns.scatterplot(data=df, x='Income', y='Age')
```

```python
from sklearn.preprocessing import StandardScaler

X = df[['Income', 'Age']]
Xs = StandardScaler().fit_transform(X)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
inertias = []
sil_scores = []

for k in range(2, 10):
    km = KMeans(n_clusters=k)
    km.fit(Xs)
    inertias.append(km.inertia_)
    sil_scores.append(silhouette_score(Xs, km.labels_))
```

```python
sns.lineplot(x=range(2, 10), y=inertias)
```

```python
sns.lineplot(x=range(2, 10), y=sil_scores)
```

```python
k = 5

km = KMeans(n_clusters=k)
y = km.fit_predict(Xs)
```

```python
plt.scatter(*Xs.T, s=10, c=y)
```

```python
df['cluster'] = y
```
