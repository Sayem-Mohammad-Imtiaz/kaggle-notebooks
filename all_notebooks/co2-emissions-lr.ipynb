```python
test_size = int(len(df)*0.3)
train_set = df[:-test_size]
test_set = df[-test_size:]
```

```python
train_set_x = train_set['Fuel Consumption City (L/100 km)']
train_set_x = train_set_x[:, np.newaxis]
train_set_y = train_set['CO2 Emissions(g/km)']
train_set_y = train_set_y[:, np.newaxis]

test_set_x = test_set['Fuel Consumption City (L/100 km)']
test_set_x = test_set_x[:, np.newaxis]
test_set_y = test_set['CO2 Emissions(g/km)']
test_set_y = test_set_y[:, np.newaxis]
```

```python
print(train_set_x.shape)
print(train_set_y.shape)
print(test_set_x.shape)
print(test_set_y.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(train_set_x, train_set_y)
```

```python
print("Slope:", lm.coef_)
print("Intercept:", lm.intercept_)
```

```python
y_train_pred = lm.predict(train_set_x)
print(y_train_pred[:5])
```

```python
print(train_set_y[:5])
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(train_set_y, y_train_pred)
```

```python
from sklearn.metrics import r2_score
r2_train_acc = r2_score(train_set_y, y_train_pred)
print("Accuracy:", r2_train_acc*100)
```

Test set prediction

```python
y_test_pred = lm.predict(test_set_x)
print(y_test_pred[:5])
```

```python
print(test_set_y[:5])
```

```python
mean_absolute_error(test_set_y, y_test_pred)
```

```python
r2_test_acc = r2_score(test_set_y, y_test_pred)
print("Accuracy:", r2_test_acc*100)
```
