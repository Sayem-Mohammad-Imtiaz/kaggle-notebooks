# Assign Different values To Different Outlet Sizes

```python
data['Outlet_Type']=data['Outlet_Type'].replace({'Supermarket Type1':1,'Supermarket Type2':2,'Grocery Store':3, 'Supermarket Type3' :4})
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data[['Item_Weight','Item_Visibility','Item_MRP','Item_Outlet_Sales','Outlet_Establishment_Year']],data['Outlet_Type'],test_size=0.3)
```

```python
X_train
```

```python
y_train
```

# Apply Logistics Regression Model 

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=230)
model.fit(X_train,y_train)
```

# Prediction 

```python
model.predict(X_test)
```

# Model Score

```python
scoretrain = model.score(X_train, y_train)
print("The accuracy of Logistic regression on Training data is", scoretrain)

scoretest = model.score(X_test,y_test)
print("The accuracy of Logistic Regression on the Test data is", scoretest)
```

```python
y_predicted=model.predict(X_test)
print(y_predicted)
```

```python
from sklearn.metrics import classification_report
clf_report = pd.DataFrame(classification_report(y_test, y_predicted, output_dict=True))
```
