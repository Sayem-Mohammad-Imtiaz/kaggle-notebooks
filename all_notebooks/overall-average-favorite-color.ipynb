```python
import pandas as pd

def get_rgb(row):
    return tuple(int(row["Favorite color"].lstrip("#")[i : i + 2], 16) for i in (0, 2, 4))

df = pd.read_csv("../input/favorite-colors/responses.csv")

colors = df.apply(get_rgb, axis=1).tolist() # Get a list of RGB colors

# Split the RGB colors into red, green, and blue values
reds = [color[0] for color in colors]
greens = [color[1] for color in colors]
blues = [color[2] for color in colors]
```

Now, when it comes to averaging RGB colors, we need to square each set of values, find their averages, and get the square roots. The reasoning behing this is described well by Kevin Simper [here](https://kevinsimper.medium.com/how-to-average-rgb-colors-together-6cd3ef1ff1e5).

```python
from math import sqrt

def average(values):
    # Return the square root of the average of the squares
    return sqrt(sum(value**2 for value in values) / len(values))

# Put the averages of the three lists together, rounding to the nearest integer
avg_color = (round(average(reds)), round(average(greens)), round(average(blues)))

# Convert to hex format
avg_color = '#%02x%02x%02x' % avg_color

avg_color
```

Show the color visually with a simple polygon fill without axes.

```python
import matplotlib.pyplot as plt

plt.axis("off")

plt.fill([0, 0, 1, 1], [0, 1, 1, 0], avg_color)

print("The average color is shown below.")
plt.show()
```
