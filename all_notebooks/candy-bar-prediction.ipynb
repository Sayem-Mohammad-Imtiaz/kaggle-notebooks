```python
x = df.drop(['hard', 'bar', 'sugarpercent', 'winpercent'], axis = 1)
y = df['bar']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
LRmodel = LogisticRegression()
LRmodel.fit(x_train, y_train)
y_pred = LRmodel.predict(x_test)
print("accuracy score using Logistic model is : ", accuracy_score(y_test, y_pred))
print("confusion matrix using Logistic model is : ", confusion_matrix(y_test, y_pred))
print("classification report using Logistic model", classification_report(y_test, y_pred))
```

```python
treeModel = DecisionTreeClassifier()
treeModel.fit(x_train, y_train)
y_prediction = treeModel.predict(x_test)

print("accuracy score using Tree model is : ", accuracy_score(y_test, y_prediction))
print("confusion matrix using Tree model is : ", confusion_matrix(y_test, y_prediction))
print("classification report using Tree model", classification_report(y_test, y_prediction))
```

```python
KNNModel = KNeighborsClassifier(n_neighbors = 13)
KNNModel.fit(x_train, y_train)
predictions = KNNModel.predict(x_test)

print("accuracy score using KNN model is : ", accuracy_score(y_test, predictions))
print("confusion matrix using KNN model is : ", confusion_matrix(y_test, predictions))
print("classification report using KNN model", classification_report(y_test, predictions))
```

```python
SVMmodel = SVC()
SVMmodel.fit(x_train, y_train)
predictions_y = SVMmodel.predict(x_test)

print("accuracy score using SVM model is : ", accuracy_score(y_test, predictions_y))
print("confusion matrix using SVM model is : ", confusion_matrix(y_test, predictions_y))
print("classification report using SVM model", classification_report(y_test, predictions_y))
```
