# Anime Name Wordcloud

```python
from os import path, getcwd
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator
import requests
url='https://i.pinimg.com/originals/dc/ab/b7/dcabb7fbb2f763d680d20a3d228cc6f9.jpg'
im = requests.get(url, stream=True).raw
mask = np.array(Image.open(im))
wc = WC(background_color="white", max_words=1000, mask=mask,
               max_font_size=90, random_state=42)
wc.generate(' '.join(data.Anime_Name))
# create coloring from image
image_colors = ImageColorGenerator(mask)
plt.figure(figsize=[15,15])
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
_=plt.show()
```
