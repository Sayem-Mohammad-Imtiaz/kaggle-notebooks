```python
# there are some outliers in the dataset lets remove them
for columns in ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']:
       df=df[df[columns]< df[columns].quantile(0.90)]
```

```python
plt.figure(figsize=(20,20))
pltnum=1
for columns in df:
    if pltnum <= 12:
        ax=plt.subplot(4,3,pltnum)
        sns.boxplot(x=df[columns])
    pltnum=pltnum+1
```

Doing Eda with Dependent and Independent varables

```python
for columns in ['fixed acidity', 'volatile acidity', 'citric acid',
       'residual sugar', 'chlorides', 'free sulfur dioxide',
       'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']:


        fig,ax=plt.subplots(1,3,figsize=(20,5))
        sns.barplot(data=df,x='quality',y=df[columns],ax=ax[0])
        sns.lineplot(data=df,x='quality',y=df[columns],ax=ax[1])
        sns.violinplot(data=df,x='quality',y=df[columns],ax=ax[2])
```

From the above bivariate analysis
we can say that there id some strong relation between below listed variables and quality
- Volatile acidity
 - Citric acid

 - Density
 - Alcohol

```python
x=df.drop(['quality'],axis=1)
y=df['quality']
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test,y_train,y_test =train_test_split(x,y,test_size=.1,random_state=3)
from sklearn.ensemble import RandomForestClassifier
model= RandomForestClassifier()
```

```python
# fitting the data into the model
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
pred
```

```python
from sklearn.metrics import accuracy_score
# now chcking the the accuracy of the model
acc=accuracy_score(pred,y_test)
```

```python
print("The accuracy of the model ",acc)
```
