Columns Director, cast,date added  has maximum number of null values and these column does not have a huge impact in data analysng so these column can be removed 

```python

del df['director']
del df['cast']
del df['date_added']
```

We even have maximum null values in the Country Column, so instead of dropping the whole column we can just replace the null values in that column with United States as there are many shows are aired in that country and Netflix itself is created in that country.

```python
df['country'].replace(np.nan, 'United States',inplace  = True)
```

```python
#value counts of rating column
df['rating'].value_counts()
```

There are very less null values in the rating column so we can replace that with the most rated one i.e TV-MA

```python
df['rating'].replace(np.nan, 'TV-MA',inplace  = True)
```

we have handled all the null values, now you can check if all the null values are removed

```python
df.isnull().sum()
```
