```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load dataset
data = pd.read_csv('/kaggle/input/simple-house-price-prediction/house_price.csv')
data.head()
```

## Feature Normalization
By looking at the values, we can see that areas are very large than the number of rooms. When features differ by orders of magnitude, performing feature scaling can make gradient descent converge much more quickly

```python
# normalization
data=np.array(data,dtype=float)
for i in range(0,data.shape[1]-1):
    data[:,i]=(data[:,i]-np.mean(data[:,i]))/np.std(data[:,i])
```

```python
x=data[:,:-1]
y=data[:,-1]
y=np.reshape(y,(47,1))
ones=np.ones((x.shape[0],1))
x=np.hstack((ones,x))
```

## hypothesis function

```python
# function to return hypothesis
def h(x,theta):
    return np.matmul(x,theta)
```

## Cost Function

```python
# function to return cost function
def cost_f(x,y,theta):
    np.square(h(x,theta)-y)/(2*x.shape[0])
```

```python
theta=np.zeros((x.shape[1],1))
```

## Finding the optimal parameters using Gradient Descent

```python
learning_rate=0.01
epoche=400
j_all=[]
for i in range(epoche):
    h_x = h(x, theta)
    cost_ = (1/x.shape[0])*(x.T@(h_x - y))
    theta = theta - (learning_rate)*cost_
    j_all.append(cost_f(x,y,theta))
theta
```

```python
y_predict = np.dot(x,theta)
```

## visualize predicted data

```python
plt.scatter(x[:,1],y)
plt.scatter(x[:,1],y_predict)
```

## Model Eveluation

```python
from sklearn import metrics
print(metrics.mean_absolute_error(y, y_predict))
print(metrics.mean_squared_error(y, y_predict))
print(np.sqrt(metrics.mean_squared_error(y, y_predict)))
```
