**Adjusted Offensive and Defensive Efficiency(ADJOE/ADJDE) are two of the most important statistics to look at in order to determine the strength of a team. **

Adjusted offensive efficiency is an estimate of the points the team would score per 100 possessions against the average NCAA defense, 

Conversely, adjusted defensive efficiency is an estimate of how many points the team would allow to the average NCAA offense over the course of 100 possessions.
> I wonder whether offense or defense is more important for teams.


**First, letâ€™s find out what the average adjusted offensive and defensive efficiency rating is throughout the 2015-2019 NCAA basketball seasons.  **
Save these values as adj_off and adj_def, and output them separated by a comma. 

*(Hint: the dataset is comprised of data from the 2015-2019 seasons, the columns containing offensive and defensive efficiency are labeled ADJOE and ADJDE, respectively)*

```python
avg_off = team_stats['ADJOE'].mean()
avg_def = team_stats['ADJDE'].mean()
print(avg_off,  avg_def, sep = ',')
```

**Interesting, can you think of why the average ADJOE and ADJDE is the same?  Do you predict that the average adjusted offensive and defensive efficiency ratings among championship teams will be equal to one another as it was among the entirety of the teams?
**

(Hint: think about the definitions of adjusted offensive and defensive efficiency 

**Which side of the ball (offense or defense), is more important to championship teams?  **

*(Hint: Do championship teams gain differ more from average in their defensive or offensive efficiency?) *

```python

team_stats[team_stats['POSTSEASON'] == 'Champions']['ADJOE'].mean() -avg_off
```

```python
avg_def - team_stats[team_stats['POSTSEASON'] == 'Champions']['ADJDE'].mean()
```

 So it looks like chamionship teams differed more from the average on offense than they did on defense.  
> And they say that defense wins championships.

Let's analyze the importance of offense a little bit more and find out if offense is the key to winning championships, or if teams need a combination of offense and defense to win

If offense is directly linked to winning, then the team with the highest adjusted offensive efficiency rating should also have won a championship.  Let's find out how that team did in the postseason

```python
team_stats['ADJOE'].idxmax()
team_stats.loc[1]['POSTSEASON']
```

This is an interesting result.  The fact that Wisconsin finished 2ND in 2015 with the highest ADJOE tells us that there is more to winning than just offense, but the fact that Wisconsin finished 2ND out of 64 teams suggests that it is likely that there is a relationship between offense and winning

**Lets put this theory to the test!
**

Lets see if the average adjusted offensive efficiency is indicative of winning.  We can check how the offensive efficiency compares to the number of games played (The number of games played depends on how far teams go into the tournament, so teams that have played more games have won more games)

```python
games_Played= team_stats.groupby('G')['ADJOE'].mean()
print(games_Played)
```
