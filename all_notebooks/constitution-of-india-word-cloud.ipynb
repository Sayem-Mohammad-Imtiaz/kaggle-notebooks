```python
import numpy as np
import pandas as pd
import glob
import codecs
import requests
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
from PIL import Image
```

```python
df = pd.read_csv('/kaggle/input/constitution-of-india/Constitution Of India.csv')

text_array = df['Articles'].values
text_string = "".join(text_array)
text = text_string.replace("\n", " ")
```

```python

IMG = Image.open(requests.get("https://www.pngarts.com/files/3/India-Map-Transparent-Images.png", stream=True).raw)
IMG_mask = Image.new("RGB", IMG.size, (255,255,255))
IMG_mask.paste(IMG,IMG)
IMG_mask = np.array(IMG_mask)
```

```python
stopwords = set(STOPWORDS)
stopwords.add("WHEN_CASE_STOPWORDS_NEED")

wordcloud = WordCloud(width = 1200,
                      height = 1300,
                      max_words = 150,
                      stopwords = stopwords,
                      mask=IMG_mask,
                      background_color = "white",
                      contour_width=5,
                      contour_color='steelblue',
                      max_font_size = 200,
                      min_word_length= 2,
                      random_state=40
                     ).generate(text)

fig = plt.figure(figsize=(9,10))
plt.imshow(wordcloud)
plt.axis('off')
```
