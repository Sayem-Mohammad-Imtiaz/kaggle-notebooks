# Insights from the data

```python
sns.relplot(x="Decision", y="GPA", data=df)
```

```python
sns.relplot(x="Decision", y="GMAT", data=df)
```

```python
sns.relplot(x="GPA", y="GMAT",hue = "Decision",  data=df)
```

you can clearly see that the higher both GPA and GMAT the more admit, the lower both GPA and GMAT the more nonadmit.



we can see that points have high GPA and low GMAT  still get admit,and points with high GMAT and low GPA get nonadmit!!!

We can say from above plots that the **GPA** is crucial and the Decision for getting an admit highly is dependant on it.

```python
sns.relplot(x="GPA", y="GMAT", hue="Decision", kind="line", data=df)
```

```python
#heatmap to show correlation between data
ndf = df.copy()
ndf["Decision"] = ndf["Decision"].astype("category")
ndf["Decision"] = ndf["Decision"].cat.codes

sns.heatmap(ndf.corr()**2 ,annot=True,square=True, cbar_kws= {'orientation': 'horizontal'} )
```

You can see that the **Decision** is highly correlated with **GPA**
