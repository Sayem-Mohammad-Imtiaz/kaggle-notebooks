```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
x = np.random.normal(size=1000)
#sns.distplot(x,bins=20,kde=False,hist=False)
sns.distplot(x,bins=20)
```

```python
x = np.random.poisson(lam=1,size=100)
y = np.random.poisson(lam=5,size=100)
z = np.random.poisson(lam=10,size=100)
sns.distplot(x,label='lamda 1')
sns.distplot(y,label='lamda 5')
sns.distplot(z,label='lamda 10')
plt.legend()
```

```python
path = '../input/seaborn-practice/iris.csv'
iris = pd.read_csv(path)

# Show the results of a linear regression within each dataset
sns.lmplot(x='sepal_length', y="petal_length",data=iris)
```

```python
sns.relplot(x='sepal_length', y="petal_length",hue='species',data=iris)
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris)
```

```python
sns.pairplot(iris,hue="species",palette="coolwarm")
```

```python
path = '../input/seaborn-practice/tips.csv'
tips = pd.read_csv(path)
sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips)
```

```python
sns.boxplot(x="day", y="total_bill",hue="smoker",data=tips)

# Add in points to show each observation
#sns.stripplot(x="day", y="total_bill",hue="smoker",data=tips)
```

```python
smoke_times = tips[['sex','smoker']].groupby(['sex']).count()
smoke_times.plot.pie(subplots=True,autopct="%.1f%%",legend=False)

day_times = tips[['total_bill','day']].groupby(['day']).sum()
day_times.plot.pie(subplots=True,autopct="%.1f%%",legend=False)
```

```python
# Plot the responses for different events and regions
path = '../input/seaborn-practice/fmri.csv'
fmri = pd.read_csv(path)
sns.lineplot(x="timepoint", y="signal",hue="region", style="event",data=fmri)
```

```python
path = '../input/seaborn-practice/flights.csv'
flight = pd.read_csv(path)

# Transform data into tabel form
flight = flight.pivot("month", "year", "passengers")

sns.heatmap(flight,linewidths=.5)
```
