```python
df =  pd.read_csv('/kaggle/input/sp-500-companies/constituents-financials.csv')
df.head()
```

```python
df.Sector.value_counts()
```

```python
def sector_distributions(Sector):
    fig, ax = plt.subplots(1, 7, figsize=(15, 6))
    sns.histplot(df[df['Sector'] == Sector]['Price/Earnings'], ax=ax[0])
    sns.histplot(df[df['Sector'] == Sector]['Dividend Yield'], ax=ax[1])
    sns.histplot(df[df['Sector'] == Sector]['Earnings/Share'], ax=ax[2])
    sns.histplot(df[df['Sector'] == Sector]['Market Cap'], ax=ax[3])
    sns.histplot(df[df['Sector'] == Sector]['EBITDA'], ax=ax[4])
    sns.histplot(df[df['Sector'] == Sector]['Price/Sales'], ax=ax[5])
    sns.histplot(df[df['Sector'] == Sector]['Price/Book'], ax=ax[6])
    fig.show()
```

```python
sector_distributions('Consumer Discretionary')
```

```python
sector_distributions('Information Technology')
```
