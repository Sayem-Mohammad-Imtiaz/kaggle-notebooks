```python
#Code by Siti K https://www.kaggle.com/khotijahs1/2020-indonesia-university-rank/comments

#10 Nonsurgical Procedures performed by Plastic Surgeons
top_procedure = df.sort_values(by='percent', ascending=False)[:10]
figure = plt.figure(figsize=(10,6))
sns.barplot(y=top_procedure.nonsurgical_procedures, x=top_procedure.percent, color='g')
plt.xticks()
plt.xlabel('Percent')
plt.ylabel('Nonsurgical Procedures')
plt.title('Nonsurgical Procedures performed by Plastic Surgeons')
plt.show;
```

```python
from collections import Counter
procedure_data = df['nonsurgical_procedures']
procedure_count = pd.Series(dict(Counter(','.join(procedure_data).replace(' ,',',').replace(
    ', ',',').split(',')))).sort_values(ascending=False)
```

```python
top10procedure = procedure_count.head(10)
```

```python
#Code by Mohammad Inran Shaikh https://www.kaggle.com/shikhnu/data-analysis-and-visualization-netflix-data
from matplotlib import gridspec

fig = plt.figure(figsize=(20, 7))
gs = gridspec.GridSpec(nrows=1, ncols=2, height_ratios=[6], width_ratios=[10, 5])

ax = plt.subplot(gs[0])
sns.barplot(top10procedure.index, top10procedure, ax=ax, palette="RdGy")
ax.set_xticklabels(top10procedure.index, rotation='90')
ax.set_title('Nonsurgical Procedures', fontsize=15, fontweight='bold')

ax2 = plt.subplot(gs[1])
ax2.pie(top10procedure, labels=top10procedure.index, shadow=True, startangle=0, colors=sns.color_palette("RdGy", n_colors=20),
       autopct='%1.2f%%')
ax2.axis('equal') 

plt.show()
```

```python
#word cloud
from wordcloud import WordCloud, ImageColorGenerator
text = " ".join(str(each) for each in df.nonsurgical_procedures)
# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=200,colormap='Set2', background_color="black").generate(text)
plt.figure(figsize=(10,6))
plt.figure(figsize=(15,10))
# Display the generated image:
plt.imshow(wordcloud, interpolation='Bilinear')
plt.axis("off")
plt.figure(1,figsize=(12, 12))
plt.show()
```
