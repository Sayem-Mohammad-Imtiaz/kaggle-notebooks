```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

for dirname, _, filenames in os.walk('../input/abates'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#Previsão de abates anual

abates = pd.read_csv('../input/ipeadata/abates_anual.csv')
abates
```

```python

abates.columns = ['ano', 'abates_bovinos','vazio']
abates.pop("vazio")

abates
```

```python
abates = abates.dropna()
print("DataFrame after removing rows with NaN value in All Columns:")
#print(abates.to_string())
```

```python
x = abates['ano'].values
y = abates['abates_bovinos'].values
```

```python
plt.scatter(x,y,label="x(y)")
plt.xlabel("ano")
plt.ylabel("abates_bovinos")
plt.legend()
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()

x = x.reshape(-1, 1)
modelo.fit(x,y)
acuracia = modelo.score(x,y)
print("Acuracia = %.2f" %acuracia)
coef_angular = modelo.coef_[0]
coef_linear_intercerpt = modelo.intercept_
print("Coeficiênte Angular = %.2f" %coef_angular)
print("Coeficiênte Linear = %.2f" %coef_linear_intercerpt)
```

```python
reta = coef_angular * x + coef_linear_intercerpt
plt.scatter(x,y,label="x(y)")
plt.plot(x,reta,label="Ajuste Linear", color='red')
plt.title("Regressão Linear Simples")
plt.xlabel("ano")
plt.ylabel("abates_bovinos")
plt.legend()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mae = mean_absolute_error(y, reta)
rmse = np.sqrt(mean_squared_error(y, reta))

print("MAE = %.2f" %mae)
print("RMSE = %.2f" %rmse)
```

```python
previsao = modelo.predict(np.array([2021]).reshape((-1, 1)))
print("Previsão de abates: %.3f,00 milhões" %previsao)

previsao2 = coef_linear_intercerpt + coef_angular * 2021
print("Previsão 2 de abates: %.3f,00 milhões" %previsao2)
```

```python
'''Este modelo dividindo a base em teste e treino faz o resultado final variar'''


# Definindo as colunas de x e de y para a regressão linear simples
x = pd.DataFrame(abates.ano)
y = abates.abates_bovinos

# Divide o dataset entre treino (70%) e teste (30%), sendo ano a variavel independente e abates_bovinos a variavel dependente. Welcome to Cross-validation!!!
from sklearn.model_selection import train_test_split
xtreino, xteste, ytreino, yteste = train_test_split(x, y, test_size=0.3)
xtreino.shape, xteste.shape, ytreino.shape, yteste.shape



# Gera o modelo de regressão linear (lm) no dataset de treino
from sklearn.linear_model import LinearRegression
modelolm = LinearRegression().fit(xtreino, ytreino)

# Faz as previsoes no dataset de teste
previsoes = modelolm.predict(xteste)
previsoes[0:9]

# Validação e Performance do Modelo. Qual foi o score (R2)?
modelolm.score(xteste, yteste)

# Qual foi o Intercept?
modelolm.intercept_

# Qual foi o Slope?
modelolm.coef_

# Avaliação dos Erros do Modelo - MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(yteste, previsoes)
mae

# Descobrir a quantidade de demissoes de apenas um valor unico de admissoes
#print(modelolm.predict(np.array([2021]).reshape((-1, 1))))
resultado = modelolm.predict(np.array([2021]).reshape((-1, 1)))
print("Previsão de abates: %.3f,00 milhões" %resultado)
```
