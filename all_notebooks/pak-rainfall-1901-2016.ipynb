```python

g1 = sb.distplot(np.log(df['Rainfall - (MM)'] + 1))
g1.set_title("Rainfall - (MM) Distribution", fontsize=16);
```

```python
# ADF Test
result = adfuller(df['Rainfall - (MM)'], autolag='AIC')
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')
```

```python
# KPSS Test
result = kpss(df['Rainfall - (MM)'], regression='c')
print('\nKPSS Statistic: %f' % result[0])
print('p-value: %f' % result[1])
for key, value in result[3].items():
    print('Critial Values:')
    print(f'   {key}, {value}')
```

```python
result = seasonal_decompose(df['Rainfall - (MM)'], model='additive', period=12)
result.plot();
pyplt.show();
```

```python
result = seasonal_decompose(df['Rainfall - (MM)'], model='multiplicative', period=12)
result.plot();
pyplt.show();
```
