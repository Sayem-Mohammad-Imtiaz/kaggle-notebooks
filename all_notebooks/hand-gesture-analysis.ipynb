```python
colnames = [f"mr{i}s{j}" for i in range(1, 9) for j in range(1, 9)] + ["class"]
rock = pd.read_csv("../input/emg-4/0.csv", names=colnames)
scissors = pd.read_csv("../input/emg-4/1.csv", names=colnames)
paper = pd.read_csv("../input/emg-4/2.csv", names=colnames)
ok = pd.read_csv("../input/emg-4/3.csv", names=colnames)
full = pd.concat([rock, scissors, paper, ok])

rock.shape, scissors.shape, paper.shape, ok.shape, full.shape
```

```python
full.describe()
```

```python
df = pd.DataFrame()

for i in range(1, 9):
    columns = [f"mr{j}s{i}" for j in range(1, 9)]
    df[f"s{i}"] = full[columns].values.reshape(-1)
    
df['class'] = full['class'].repeat(8).values
print(df.shape)
df.head()
```
