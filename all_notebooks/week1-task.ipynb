```python
#importing libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import style
%matplotlib inline
```

```python
#reading data
df= pd.read_csv("../input/up-crime-data-year-2019/MR data - Compiled Data Set.csv")
df.head()
```

```python
#dropping unnecessary columns 
df=df.drop(["Title","Text"],axis=1)
```

```python
#filling numerically missing data with 0
df=df.fillna(0)
```

```python
#storing the titles of all the columns in col
col=df.columns
```

```python
#renaming columns as the names provided would make data analysis pretty lengthy
df.columns=[i for i in range(0,len(df.columns))]
```

```python
d={0:0,'Property Disputes':1,'Family Dispute':2,'Petty Quarrels':3,'Money Disputes':4, 'Personal Vendetta':5, 'Love Affairs':6, 'Casteism':7,'Unknown reasons':8}
```

```python
df[1]=df[1].apply(lambda x :d[x])
```

```python
#seeing how many reasons are there at the same time for the crime against women
for i in range(0,len(df)):
    if df.iloc[i,8]!=0:
        print(df.iloc[i,8])
```

```python
#creating extra columns to store the different crime reasons provided in the same crimes against women column into different columns
df[11]=0
df[12]=0
```

```python
#splitting crime against women column into three different columns
for i in range(0,len(df)):
    if ',' in str(df.iloc[i,8]):
        c=str(df.iloc[i,8]).split(',')
        df.iloc[i,8]=int(c[0])
        df.iloc[i,11]=int(c[1])
        if len(c)>2:
            df.iloc[i,12]=int(c[2])
```
