```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.Stay=le.fit_transform(df['Stay'])
```

```python
cols=df.columns
for col in cols:
    print ('No. of unique categories for ', col, 'are', df[col].nunique())
```

```python
# DROP case_id and patientid since they have multiple unique rows
df.drop(['case_id','patientid'], axis=1, inplace=True)
```

```python
# convert NA values from Bed Grade column to 1
df['Bed Grade']=np.where(df['Bed Grade'].isna(),1.0,df['Bed Grade'])
```

```python
# convert NA values from City_Code_Patient column to 0
df['City_Code_Patient']=np.where (df['City_Code_Patient'].isna(),0.0,df['City_Code_Patient'])
```
