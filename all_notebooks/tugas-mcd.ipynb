**MATPLOTLIB**

```python
girl = pd.read_csv("../input/kpopdb/kpop_idols_girl_groups.csv")
girl
```

```python
girl.info()
```

```python
#Linechart

import matplotlib.pyplot as plt
  
plt.plot(girl['Name'], girl['Members'], color='blue', marker='^')
plt.title('Name Vs Members', fontsize=14) # untuk menampilkan judul
plt.xlabel('Name', fontsize=14) # untuk mencetak label di sumbu x
plt.ylabel('Members', fontsize=14) # untuk mencetak label di sumbu Y
plt.grid()
plt.show() # untuk menampilkan visualisasi
```
