# Objective

The goal of this notebook is to show the correlation of text features such as readability scores with the relative complexity rating (the "target" variable) of the CommonLit Readability challenge.

```python
import pandas as pd
import plotly.express as px
```

```python
data_path = "../input/commonlit-text-features/augmented_df.csv"
df = pd.read_csv(data_path)
```

# Data

Let's see how the data looks like:

```python
print(df.columns)
df.head()
```

```python
target_col = "target"
feature_cols = df.columns[4:]
```

# Plotting time!

```python
for feat in feature_cols:
    fig = px.scatter(df, x=feat, y=target_col)
    fig.show()
```

# Conclusion

A visual inspection shows there are some potentially useful correlations between these metrics and the target variable. This means they might be used in some models, or as a quick sanity check for out-of-sample forecasting, to ensure models don't return crazy values when seeing test data.
