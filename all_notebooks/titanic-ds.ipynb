## Random forest

```python
#splitting the data for train, and selecting the attributes on which training and testing is to be done
x_train = pd.get_dummies(titanic_train[["Pclass", "Sex", "SibSp", "Parch"]])
y_train = titanic_train["Survived"]
y_test = titanic_test["Survived"]
x_test = pd.get_dummies(titanic_test[["Pclass", "Sex", "SibSp", "Parch"]])

#creating our model
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=9, max_depth=5, random_state=1)
model = clf.fit(x_train, y_train)

#predicting the results
y_pred = clf.predict(x_test)


```

```python
from sklearn import metrics

acc=metrics.accuracy_score(y_test,y_pred,3)
print(acc)
```

```python
#pretty table
from prettytable import PrettyTable

AccuracyTable = PrettyTable(["Algorithm", "Accuracy "])
AccuracyTable.add_row(["Random forest", acc])
print(AccuracyTable)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics

clf = DecisionTreeClassifier(criterion = 'entropy')
model=clf.fit(x_train,y_train)
#prob=clf.predict_proba([[4,3.2,5,6]])

y_pred = clf.predict(x_test)
accuracy_score_clf = accuracy_score(y_test,y_pred,3)
print(accuracy_score_clf)
```

```python
AccuracyTable.add_row(["Decision Tree", accuracy_score_clf])
print(AccuracyTable)
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=7)
model.fit(x_train,y_train)
predicted= model.predict(x_test)
acc = metrics.accuracy_score(y_test, predicted)
print("Accuracy:",acc)
```

```python
AccuracyTable.add_row(["KNN", acc])
print(AccuracyTable)
```

## Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB

#model = GaussianNB()

model = MultinomialNB()
model.fit(x_train,y_train,3)
predicted = model.predict(x_test)
acc_NB = metrics.accuracy_score(y_test, predicted)
print("Accuracy:", acc_NB)
```

```python
AccuracyTable.add_row(["Naive bayes ", acc_NB])
print(AccuracyTable)
```

## SVM

```python
from sklearn import svm
clf = svm.SVC(kernel='poly')
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
acc_SVM = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:",acc_SVM)
```

```python
AccuracyTable.add_row(["SVM ", acc_SVM])
print(AccuracyTable)
```
