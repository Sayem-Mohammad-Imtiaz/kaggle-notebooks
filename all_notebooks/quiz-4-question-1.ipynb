<h1>2. Selecting and Summarizing Data<h1>

a. Let's begin by displaying all the fights in Los Angeles

**Explanation: This teaches us how to narrow into fights in a city we have interest in**

```python
data[data.location == 'Los Angeles, California, USA']
```

b. Next let's see what the mean reach of all the fighters is

**Explanation: We will do this by calculating the average reach of red fighters and blue fighters.
This will help us gain an idea of whether a potential fighter has what it takes to stick in the UFC. If they have a longer reach than average, that might give them an advantage.**

```python
(data['R_Reach_cms'].mean() + data['B_Reach_cms'].mean())/2
```

c. Now, let's find out what the minimum reach for a fighter is

**Explanation: In case a potential fighter doesn't reach the average, what is the minimum reach he would need to be able to compete in the UFC? This information is useful for learning the standards required to fight**

```python
data[['R_Reach_cms', 'B_Reach_cms']].min()
```

d. Next, let's take a look at mean reach, grouped by fighting style

**Explanation: To do this, we have to make sure to group by fighting stance for both fighters and select the reach of both fighters.
This could help us find out if fighters with more reach favor certain fighting styles, it will also allow us to see different combintations of fighting styles**

```python
data.groupby(['R_Stance','B_Stance'])[['R_Reach_cms', 'B_Reach_cms']].mean()
```

<h1> 3. Plotting<h1>

a. Let's go ahead and find out which cities have the most fights and plot them in a pie chart

**Explanation: To do this, we have to first clean the data by removing all fights with cities set to 'None', next lets group by location, count the number of fights by using size, sort in descending order, then take the cities with the five fights. This is useful, because it would help us learn where the majority of UFC's business is located, perhaps we should focus our attention on a city other than Los Angeles**

```python
cities = data.replace('None', np.NaN).groupby('location').size().sort_values(ascending=False).iloc[:5]
cities.plot.pie()
```
