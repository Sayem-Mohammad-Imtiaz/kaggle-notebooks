<h2 id="reading_data">Reading the data in</h2>

```python

df = pd.read_csv('../input/housedata/output.csv')

# take a look at the dataset
df.head()
```

Lets select some features that we want to use for regression.

```python
feat = df[['bedrooms', 'bathrooms', 'sqft_living', 'floors', 'yr_built', 'price']]
feat.head(11)
```

Lets plot Price with respect to Square foot living:

```python
sqft = feat.sqft_living/1000
rate = feat.price/10000
plt.scatter(sqft, rate, color='green')
plt.xlabel("Sqft_living(in thousands)")
plt.ylabel("Price(in lakhs)")
plt.show()
```

#### Creating train and test dataset
Train/Test Split involves splitting the dataset into training and testing sets respectively, which are mutually exclusive. After which, you train with the training set and test with the testing set.

```python
split = np.random.rand(len(df)) < 0.8
train = feat[split]
test = feat[~split]
```

### Train data distribution

```python
X = train.sqft_living/1000
Y = train.price/10000
plt.scatter(X, Y, color='blue')
plt.xlabel("Train_sqft(in thousands)")
plt.ylabel("Train_rate(in lakhs)")
plt.show()
```

<h2 id="multiple_regression_model">Multiple Regression Model</h2>
In reality, there are multiple variables that can predict the Price. When more than one independent variable is present, the process is called multiple linear regression. For example, predicting price using Bedrooms, Bathrooms, Floors and Square foot Living. The good thing here is that Multiple linear regression is the extension of simple linear regression model.

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['bedrooms', 'bathrooms', 'sqft_living', 'floors']])
y = np.asanyarray(train[['price']])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

<h2 id="prediction">Prediction</h2>

```python
y_hat= regr.predict(test[['bedrooms', 'bathrooms', 'sqft_living', 'floors']])
x = np.asanyarray(test[['bedrooms', 'bathrooms', 'sqft_living', 'floors']])
y = np.asanyarray(test[['price']])
print("Residual sum of squares: %.2f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % regr.score(x, y))
```
