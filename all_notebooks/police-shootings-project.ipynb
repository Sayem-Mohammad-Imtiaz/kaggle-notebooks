```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv
df = pd.read_csv('../input/us-police-shootings/shootings.csv')
df.head(15)

```

```python
df['race'].value_counts().plot.pie()
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df['age'])

# set x/y labels and plot title
plt.pyplot.xlabel('age')
plt.pyplot.ylabel("count")
plt.pyplot.title('age bins')
```

```python
import seaborn as sns
sns.boxplot(x='race', y='age', data=df)
```

```python
df_group_one = df.groupby(['race'],as_index=False).mean()
df_group_one
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df['race'])

# set x/y labels and plot title
plt.pyplot.xlabel('race')
plt.pyplot.ylabel("count")
plt.pyplot.title('race bins')
```
