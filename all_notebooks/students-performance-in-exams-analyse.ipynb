```python
df['average'] = (df['math_score']+df['reading_score']+df['writing_score'])/3
def grade(average):
    if (average>=50) :return 'V'
    if (average<50) :return 'F'
df["grade"] = df.apply(lambda x : grade(x["average"]), axis=1)
df["grade_math"] = df.apply(lambda x : grade(x["math_score"]), axis=1)
df["grade_reading"] = df.apply(lambda x : grade(x["reading_score"]), axis=1)
df["grade_writing"] = df.apply(lambda x : grade(x["writing_score"]), axis=1)
```
