# Time Series

```python
df1=pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/time_series_covid_19_deaths.csv')
df2=pd.read_csv('/kaggle/input/novel-corona-virus-2019-dataset/time_series_covid_19_confirmed.csv')
df1.head()
```

```python
df1=df1.groupby(['Country/Region']).sum().drop(['Lat','Long'],axis=1)
df2=df2.groupby(['Country/Region']).sum().drop(['Lat','Long'],axis=1)
#df2.reset_index(inplace=True)
#print(df1.head())
#print(df2.head())
```

# On Same Axis

```python
df1_select=df1.loc[df1.index.isin(['India'])]
df2_select=df2.loc[df2.index.isin(['India'])]
fig, ax1 = plt.subplots(figsize=(6,8))
upto_days=90
for i,country in enumerate(df1_select.index):
    death = df1_select.loc[df1_select.index== country].values[0]
    death = death[death>1][:upto_days]
    cases=df2_select.loc[df2_select.index==country].values[0]
    cases = cases[cases>1][:upto_days]
    day = np.arange(len(death))
    #ax2 = ax1.twinx()
    ax1.plot(day,cases,'-',color='blue')
    ax1.plot(day, death, 'o-', color="red" )
    
    
#ax1.scale("log")
ax1.set_xlabel('Day Number',fontsize=14)
ax1.set_ylabel('Cases/Death',fontsize=14)
ax1.legend(['Cases','Deaths'])


plt.show()
```

# On Different Axis

```python
df1_select=df1.loc[df1.index.isin(['India'])]
df2_select=df2.loc[df2.index.isin(['India'])]
fig, ax1 = plt.subplots(figsize=(6,8))
upto_days=90
for i,country in enumerate(df1_select.index):
    death = df1_select.loc[df1_select.index== country].values[0]
    death = death[death>1][:upto_days]
    cases=df2_select.loc[df2_select.index==country].values[0]
    cases = cases[cases>1][:upto_days]
    day = np.arange(len(death))
    ax2 = ax1.twinx()
    ax1.plot(day,cases,'-',color='blue')
    ax2.plot(day, death, 'o-', color="red" )
    
    
#ax1.scale("log")
ax1.set_xlabel('Day Number',fontsize=14)
ax1.set_ylabel('Cases',fontsize=14)
ax2.set_ylabel('Death',fontsize=14)


plt.show()
```
