```python
from sklearn.cluster import KMeans
data = np.array(list(df_v_new[['Xaxis_meter','Yaxis_meter']].apply(tuple, axis=1)))
kmeans = KMeans(n_clusters=12, random_state=0).fit(data)

#Find the correcting cluster, (it must be 6 cluster)
fig, ax = plt.subplots(3,4, figsize=(12,8), sharex=True, sharey=True)
k=1
ssd=[] #(model.inertia) Sum of squared distances of samples to their clossest cluster cente
for r in range(ax.shape[0]):
    for c in range(ax.shape[1]):
        ax[r,c].scatter(data[:, 0], data[:, 1],c='black', s=20, alpha=.4)
        m=KMeans(n_clusters=k)
        m.fit(data)
        ssd.append([k, m.inertia_])
        ax[r, c].scatter(m.cluster_centers_[:, 0], m.cluster_centers_[:, 1], color = 'r', marker='D',s=50)
        ax[r, c].set_title(f'k={k}, inertia={m.inertia_:,.2f}')
        k+=1
print(ssd)
#plt.savefig('df_elbow.eps')
```

```python
#Elbow Method

#สำหรับชุดข้อมูล 2p และ 3p พอใช้ elbow แล้วค่าออกมาไม่ตรงกัับความเป็นจริง
# 2 คน elbow วัดได้ เป็น 4 เพราะ ข้อมูลที่บันทึกมีไม่มาก และ ขณะทดสอบคนไม่ได้เคลื่อนไหว จะอยู่นิ่งเป็นส่วนใหญ่
# 3 คน elbow วัดได้ เป็น 4 เพราะ ที่ตำแหน่ง (-2,7) มีวัตถุโผล่มา 2 ตัว

def elbow_plot(data, from_k=1, to_k=12):
    ssd=[]
    for k in range(from_k, to_k+1):
        m=KMeans(n_clusters=k)
        m.fit(data)
        ssd.append([k, m.inertia_])
    dd=pd.DataFrame(ssd, columns=['k', 'ssd'])
    dd['pct_chg']=dd['ssd'].pct_change() * 100
    plt.figure(figsize=(10,10))
    plt.plot(dd['k'], dd['ssd'], linestyle='--',marker='o')
    plt.title('Elbow Method')
    plt.xlabel('No. of Cluster')
    plt.ylabel('Inertia')
    plt.grid(True)
    
    
    for index, row in dd.iterrows():
        plt.text(row['k']+.02, row['ssd']+.02, f'{row["ssd"]:.2f}', fontsize=12)

ssd = elbow_plot(data, 1, 12)
#plt.savefig('df_elbow_2.eps')
```

```python
kmeans = KMeans(   n_clusters=4  , random_state=0).fit(data) #input n_clusters based on elbow method for example at 4 meters n_clusters=4

fig = plt.figure()
plt.title('X Y Scatter Plot',fontsize=20)
plt.ylabel('Longitudinal disance (meters)',fontsize=15)
plt.xlabel('Lateral distance (meters)',fontsize=15)
plt.grid(True)
ax2 = fig.add_subplot(1, 1, 1)
ax2.scatter(data[:,0],data[:,1], c = kmeans.labels_)
fig.canvas.draw()

ax2.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],c='red')
fig.canvas.draw()

#plt.savefig('df_kmean.eps',format='eps')
```
