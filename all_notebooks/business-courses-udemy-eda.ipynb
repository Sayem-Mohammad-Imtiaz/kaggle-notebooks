```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("../input/business-courses-udemy-10k-courses/udemy_output_All_Business_p1_p626.csv").dropna()
data.head()
```

**How many cources was wishlisted.**

```python
data["is_wishlisted"][ data["is_wishlisted"]==False ] = 0
data["is_wishlisted"][ data["is_wishlisted"]==True ] = 1
```

```python
if(sum(data["is_wishlisted"])):
    plt.plot( data["is_wishlisted"].unique(), data["is_wishlisted"].value_counts() )
else:
    print("Every course wasn't wishlisted")
```

**The most popular courses on the fone a number of subscribers.**

```python
data[ ["title", "num_subscribers"] ].groupby("title").mean().plot()
plt.show()
```

TOP-5 courses with highest number of subscribers.

```python
print( data[ ["title", "num_subscribers"] ].groupby("title").mean().sort_values(by="num_subscribers", ascending=False)[0:5] )
```

**Courses with highest avarage rating.**

```python
data[ ["title", "avg_rating"] ].groupby("title").mean().plot()
plt.show()
```

TOP-20 of courses with highest avarage rating.

```python
print( data[ ["title", "avg_rating"] ].groupby("title").mean().sort_values(by="avg_rating", ascending=False)[0:20] )
```

**Number of lectures.**

TOP-5 of courses with highest number of lectures.

```python
print( data[ ["title", "num_published_lectures"] ].groupby("title").mean().sort_values(by="num_published_lectures", ascending=False)[0:5] )
```

TOP-15 of courses with lowest number of lectures.

```python
print( data[ ["title", "num_published_lectures"] ].groupby("title").mean().sort_values(by="num_published_lectures", ascending=True)[0:15] )
```
