```python
# Estimate the maximum ELO score according to the odds of the bookmakers 
##
import pandas as pd, numpy as np
df = pd.read_csv('../input/atp-tour-20002016/Data.csv', encoding = 'latin1')
df
```

```python
# keep only pinnacles (PSW, PSL)
dfx=df[['Winner','Loser','PSW','PSL','Date','WRank','LRank','Round','Court','Surface','Tournament']]
dfx=dfx.dropna()
dfx
```

```python
dfx["elo1"] = np.nan
dfx["elo2"] = np.nan
```

```python
#initialise
elo={}
elom={}
tournoi={}
dati={}
for i in range(1,dfx.shape[0]-1):
  elo[dfx.iloc[i,0]] = 1200
  elo[dfx.iloc[i,1]] = 1200
elom=elo.copy()
```

```python
# elo calculation adapted according to the odds 
# it doesn't matter who wins 
m=0
nomold=''
liste = []
for i in range(0,dfx.shape[0]-50):
 g = 1/dfx.iloc[i,2]
 d = 1/dfx.iloc[i,3]
 n1 = dfx.iloc[i,0]
 n2 = dfx.iloc[i,1]
 elo1 = elo[n1]
 elo2 = elo[n2]
 pd= 1/(1+10**((elo2-elo1)/400))
 elo[n1] = elo1 + 20 * ( g * (1-pd) - d * pd)  
 elo[n2] = elo2 - 20 * ( g * (1-pd) - d * pd)
 if elo[n1]>elom[n1]:
  elom[n1]=elo[n1]
  tournoi[n1]=dfx.iloc[i,9]
  dati[n1]=dfx.iloc[i,4]
 if elo[n2]>elom[n2]:
  elom[n2]=elo[n2]
  tournoi[n2]=dfx.iloc[i,9]
  dati[n2]=dfx.iloc[i,4]
```

```python
# sort and show results
tab=sorted(elom.items(), key=lambda t: t[1],reverse=True)
for i in range(0,15):
  print(tournoi[tab[i][0]],tab[i],dati[tab[i][0]])
```
