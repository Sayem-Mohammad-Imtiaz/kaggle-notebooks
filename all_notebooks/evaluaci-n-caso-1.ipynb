```python
#Pregunta 1
#Importamos las librerías necesarias
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pyspark import SparkContext
from pyspark.sql import SparkSession 
sc = SparkContext.getOrCreate() 
spark = SparkSession(sc) 
df = spark.read.load( "../input/positivos-covid/positivos_covid.csv", format="csv", sep=";",header="true",inferSchema="true" )\
.createOrReplaceTempView("positivos_covid") #leemos y cargamos el dataset de casos COVID-19, definiendo los encabezados.
```

```python
#Pregunta 2
#Utilizando comandos SQL, realizamos el conteo de casos en el dataset y los mostramos con el encabezado "CANTIDAD_POSITIVOS"
spark.sql("select count(*) CANTIDAD_POSITIVOS from positivos_covid").show() 
```

```python
#Para la Pregunta 3, mostramos los el número de persona de sexo masculino y femenino, haciendo
#la búsqueda en la fila "SEXO", incluyendo todos los valores excepto los que tuvieran valores nulos, como ocurrió en un caso

p1=spark.sql("select SEXO,count(*) CANTIDAD_POSITIVOS from positivos_covid where sexo <>'NULL' group by sexo order by count(*) desc")
p1.show()
p1_panda=p1.toPandas()
sns.barplot(x='SEXO',y='CANTIDAD_POSITIVOS',data=p1_panda) #Luego se realizó la gráfica de frecuencia respectiva
plt.show()
```

```python
#Pregunta 4, agrupamos por departamento la cantidad de positivo utilizando nuevamente una "query" de SQL
p2=spark.sql("select DEPARTAMENTO,count(*) CANTIDAD_POSITIVOS from positivos_covid group by departamento order by count(*) desc")
p2.show(30) #Mostramos las primeras 30 filas
p2_panda=p2.toPandas()
sns.barplot(y='DEPARTAMENTO', x='CANTIDAD_POSITIVOS', capsize = 0.05, data=p2_panda) #Realizamos el gráfico de frecuencias respectivo
plt.show() #Mostramos la gráfica
```

```python
#Pregunta 5: Con SQL, en la columna "metododx" extraemos la cantidad de positivos según su prueba diagnóstica
p3=spark.sql("select metododx TIPO_PRUEBA,count(*) CANTIDAD_POSITIVOS from positivos_covid group by metododx order by count(*) desc")
p3.show()
p3_panda=p3.toPandas()
sns.barplot(x='TIPO_PRUEBA', y='CANTIDAD_POSITIVOS', capsize = 0.05, data=p3_panda) #Realizamos la gráfica de barra respectiva
plt.show()
```

```python
#Para la Pregunta 6, realizamos la segmentación por edad
p4=spark.sql("select EDAD,count(*) CANTIDAD_POSITIVOS from positivos_covid group by edad order by count(*) desc")
p4.show() #Mostramos la tabla
p4_panda=p4.toPandas()
fig, ax = plt.subplots(figsize = (20,10))
sns.barplot(x='EDAD', y='CANTIDAD_POSITIVOS', capsize = 0.01, data=p4_panda) #Presentamos la gráfica
plt.show()
```
