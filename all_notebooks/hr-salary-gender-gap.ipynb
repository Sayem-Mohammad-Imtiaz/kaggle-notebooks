```python
#Let's drop "Non Revealed" gender and "CEO" out of dataframe, because it's a noise in data
final_df = df[~df['Gender'].str.contains('Not Revealed') & ~df['Department'].str.contains('CEO')]
final_df.head(5)
```

### Analysis

So we have the same quantity of males and females in this dataset. To find out the gender salary gap we have to find and show mean and median of Annual salary grouped by gender.

```python
final_df.pivot_table('Annual Salary', 'Gender', aggfunc='mean')
```

```python
final_df.pivot_table('Annual Salary', 'Gender', aggfunc='mean').plot(kind='barh', figsize=(10,5), subplots=True)
plt.show()
#The mean value of salary is equal
```

```python
final_df.pivot_table('Annual Salary', 'Gender', aggfunc='median')
```

```python
final_df.pivot_table('Annual Salary', 'Gender', aggfunc='median').plot(kind='barh', figsize=(10,5), subplots=True)
plt.show()
#Female have more median of Annual Salary
```

In the company as a whole here is equal the mean salary by both of genders, but median value female have more than male

```python
final_df.pivot_table('Annual Salary', 'Gender', 'Department', aggfunc='mean')
```

```python
final_df.pivot_table('Annual Salary', 'Gender', 'Department', aggfunc='mean', margins=True).plot(kind='barh', figsize=(26,19), subplots=True)
plt.show()
```

As we can see at this plot the mean salary of male more than mean salary of female in **50%** of company departments, but by the mean value by all of departments female are lead.

```python
final_df.pivot_table('Annual Salary', 'Gender', 'Department', aggfunc='median')
```

```python
final_df.pivot_table('Annual Salary', 'Gender', 'Department', aggfunc='median', margins=True).plot(kind='barh', figsize=(26,19), subplots=True)
plt.show()
```

The medain salary of male more than median salary of female in **40%** of company departments. The bigest median value have females.
