```python
import pandas as pd
import numpy as np

import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Dropout
from keras.layers import LSTM
from tensorflow.keras.utils import to_categorical
```

```python
df = pd.read_csv('../input/riemann-hypothesis-1-million-prize/prime_number.csv')
df.head()
```

## We only need Prime numbers

## Create time dataset

```python
a = []
for i in range(0,len(df)):
    a.append(df.prime[i:i+15].values)
```

```python
b = []
for i in range(0,len(df)):
    try:
        b.append(df.prime[i+15])
    except:
        print(i)
```

```python
len(a)
```

```python
len(b)
```

```python
a1 = np.array(a[:148918],dtype=np.float32)
b1 = pd.Series(np.array(b[:148918],dtype=np.float32))
```

```python
a1.shape
```

```python
a1 = np.expand_dims(a1, -1)
```

```python
a1.shape
```

```python
model = Sequential()

model.add(LSTM(1024, input_shape=(15, 1),return_sequences=True))
model.add(LSTM(512, return_sequences=True,dropout=0.2))
model.add(LSTM(256, return_sequences=True,dropout=0.2))
model.add(LSTM(128, return_sequences=False))
model.add(Dense(1))
model.compile(loss='mse', optimizer=keras.optimizers.Adadelta())
model.fit(a1[:100], b1[:100], epochs=10, batch_size=10)
```
