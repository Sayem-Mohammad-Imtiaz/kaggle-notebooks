# 1. Data load

```python
dataset = image_dataset_from_directory('/kaggle/input/eurosat-dataset/EuroSAT',labels = "inferred", label_mode = 'categorical', image_size = (64, 64))
```

```python
path = '/kaggle/input/eurosat-dataset/EuroSAT/'

sample1df = pd.read_csv(path + 'train.csv')
sample2df = pd.read_csv(path + 'validation.csv')
sample3df = pd.read_csv(path + 'test.csv')

s1fname = sample1df.Filename
s2fname = sample2df.Filename
s3fname = sample3df.Filename

s1label = sample1df.Label
s2label = sample2df.Label
s3label = sample3df.Label
```

```python
image1 = []
for fname in s1fname:
    Image = image.load_img(path + fname)  
    image_arr = image.img_to_array(Image)
    image1.append(image_arr)
```

```python
image2 = []
for fname in s2fname:
    Image = image.load_img(path + fname)  
    image_arr = image.img_to_array(Image)
    image2.append(image_arr)
```

```python
image3 = []
for fname in s3fname:
    Image = image.load_img(path + fname)  
    image_arr = image.img_to_array(Image)
    image3.append(image_arr)
```

# 2.Data Preprocessing

```python
#data 정리 
x = image1 + image2
y = np.array(list(s1label) + list(s2label))
xtest = image3
ytest = np.array(s3label)

#normalization 
x = np.array(x) / 255.
xtest = np.array(xtest) / 255.

#one-hot encoding 
y = to_categorical(y, 10)
ytest= to_categorical(ytest, 10)


print(x.shape)
print(y.shape)
print(xtest.shape)
print(ytest.shape)
```

# 3. Holdout Validation 

```python
import random 
import tensorflow as tf

def seed_everything(seed: int = 42):
    random.seed(seed)
    np.random.seed(seed)
    os.environ["PYTHONHASHSEED"] = str(seed)
    tf.random.set_seed(seed)
    
seed_everything()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xval, ytrain, yval = train_test_split(x, y, test_size = 0.22, shuffle = True, stratify = y)
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xval.shape)
print(yval.shape)
```

# 4. Data Augmentation 

```python
augGen = ImageDataGenerator(rotation_range = 15, width_shift_range = 0.2, height_shift_range = 0.2)

xtrain_generator = augGen.flow(xtrain, ytrain, batch_size = 32, save_format = 'jpeg')
```

## 1) efficientB0net

```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
efficientb0 = EfficientNetB0(weights = 'imagenet', include_top = False, input_shape = (64, 64, 3))
```

```python
print(len(efficientb0.layers))
```

```python
fine_tune_at = len(efficientb0.layers) - 5

for layer in efficientb0.layers[:fine_tune_at]:
    layer.trainable = False
```

```python
x = efficientb0.output
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense,MaxPooling2D, Flatten, Conv2D, Dropout
x = MaxPooling2D(pool_size = (2,2))(x)
x = Flatten()(x)
x = Dropout(0.1)(x)
x = Dense(64, activation = 'relu')(x)
x = Dropout(0.1)(x)
x = Dense(10, activation = 'softmax')(x)

efficientb0_model = Model(inputs = efficientb0.input, outputs = x)
```

```python
import os
checkpoint_path = "efficientb0_real/cp-{epoch:04d}.ckpt"
checkpoint_dir = os.path.dirname(checkpoint_path)
cp_callback = ModelCheckpoint(checkpoint_path, verbose = 1, save_weights_only = True, period = 1)
```

```python
efficientb0_model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
history = efficientb0_model.fit(xtrain_generator, validation_data = (xval, yval), epochs = 20, callbacks = [cp_callback])
```
