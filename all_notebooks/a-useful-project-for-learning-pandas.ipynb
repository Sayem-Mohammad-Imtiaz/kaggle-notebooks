**4- Finding the column which has the most missing data**

```python
number_null = df.isnull().sum().sum()
print('There are {} missing data in the dataset'.format(number_null))
```

```python
max_null = df.isnull().sum().idxmax()
print('The column which has the most missing data is {}'.format(max_null))
```

**5- Deleting the "company" column from dataset**

```python
df = df.drop('company', axis=1)
df.head()
```

**6- Name 5 counteries which have the most passengers in the hotel**

```python
df['total_passengers'] = (df['adults'] + df['children'] + df['babies']) - df['is_canceled'];

df[['country','adults','children','babies','is_canceled','total_passengers']].groupby('country').sum().nlargest(5,'total_passengers').reset_index()
```

**7- Name of the person who has the highest ADR (average daily rate)**

```python
print("{} has paid the highest ADR which is".format(df.name[df['adr'].idxmax()]),("{} USD".format(df['adr'].max())))
```

**8- Calculate the average number of ADR**

```python
print("The average number of ADR is {}".format(round(df["adr"].mean(),2)))
```

**9- Calculate the average number of nights which passengers have stayed in the hotel.**

```python
sum_stayed_nights = df['stays_in_weekend_nights'] + df['stays_in_week_nights']
print("The average number of nights which passengers have stayed in the hotel is {}.".format(round(sum_stayed_nights.mean(),2)))
```

**10- Specify name and E-mail address of people who had 5 special request.**

```python
special_req = df[df['total_of_special_requests']==5]
special_req_info = special_req[['name', 'email','total_of_special_requests']]
special_req_info
```

**11- Specify five surenames which have the most frequency.**

```python
df['name'].apply(lambda lname:lname.split()[1]).value_counts().head().to_frame('value_counts')
```

**12- Specify people who had the most children and babies.**

```python
print("{} had the most number of children which was".format(df.name[df['children'].idxmax()]),("{}.".format(df['children'].max())))
print("{} had the most number of babies which was".format(df.name[df['babies'].idxmax()]),("{}.".format(df['babies'].max())))
```

**13- Specify the prefix of telephone number which had most frequency.**

```python
df['phone-number'].apply(lambda phone_code:phone_code[:3]).value_counts().nlargest(3).to_frame('value_counts')
```
