```python
train=pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
test=pd.read_csv('/kaggle/input/digit-recognizer/test.csv') 
train.shape
```

```python
size=int(train.shape[0]*0.1) 
print(size)
X_train=train.iloc[size:,1:].values
Y_train=train.iloc[size:,:1].values 
```

```python
X_train=X_train.reshape(-1,28,28,1)
Y_train=Y_train.reshape(-1,1) 
```

```python

X_train.shape
Y_train.shape 
```

```python
 X_val=train.iloc[:size,1:].values 
Y_val=train.iloc[:size ,:1].values
```

```python

X_val=X_val.reshape(-1,28,28,1)
Y_val=Y_val.reshape(-1,1)
```

```python
print(X_val.shape) 
print(Y_val.shape)
```

```python
X_test=test.iloc[:,0:].values 
print(X_test.shape)
Y_test=test.iloc[:, :1].values
print(Y_test.shape)
```

```python
X_test=X_test.reshape(-1,28,28,1)
print(X_test.shape)
```

```python
X_train=X_train/255.0
X_val=X_val/255.0
X_test=X_test/255.0
```

```python
plt.imshow(np.resize(X_train[0],(28,28))) 
Y_train[0]
```

```python
model = models.Sequential()
```

```python

    model.add(Conv2D(32,(3,3),activation='relu',padding='same',input_shape=(28,28,1)))
    model.add(Conv2D(32,(3,3),activation='relu',padding='same',input_shape=(28,28,1)))
    model.add(MaxPool2D(pool_size=(2,2),padding='same'))
    model.add(Dropout(0.25))
    model.add(Conv2D(64,(3,3),activation='relu',padding='same',input_shape=(28,28,1)))
    model.add(Conv2D(64,(3,3),activation='relu',padding='same',input_shape=(28,28,1)))
    model.add(MaxPool2D(pool_size=(2,2),padding='same'))
    model.add(Dropout(0.25))
    model.add(Conv2D(128,(3,3),activation='relu',padding='same',input_shape=(28,28,1)))
    model.add(Conv2D(128,(3,3),activation='relu',padding='same',input_shape=(28,28,1)))
    model.add(MaxPool2D(pool_size=(2,2),padding='same'))
    model.add(Dropout(0.25)) 
    model.add(Flatten()) 
    model.add(Dense(256,activation='relu'))
    model.add(Dense(64,activation='relu'))
    model.add(Dense(10,activation="sigmoid"))
```

```python
model.summary()
```

```python
initial_lr = 0.001
loss = "sparse_categorical_crossentropy"
model.compile(Adam(lr=initial_lr), loss=loss ,metrics=['accuracy'])
```

```python
epochs = 20
batch_size = 256
history_1 = model.fit(X_train,Y_train,batch_size=batch_size,epochs=epochs,validation_data=[X_val,Y_val])
```

```python
results = model.predict(X_test)

# select the indix with the maximum probability
results = np.argmax(results,axis = 1)

results = pd.Series(results,name="Label")
```

```python
submission = pd.concat([pd.Series(range(1,28001),name = "ImageId"),results],axis = 1)

submission.to_csv("sub.csv",index=False)
```
