# Data Cleaning

```python
data['PE Ratio'].unique()[1:50]
```

**Checking how many null values are their in `PE Ratio` attribute**

```python
data[data['PE Ratio'].isnull()]
```

**Their are any null value in PE Ratio attribute , their are many null values in other attributes, so we need to fill them as well.**

**Filling null value with "others" for attributes other than 'PE Ratio'.**

```python
data[1:-2].fillna('others',inplace=True)
data['PE Ratio'].unique()
```

**The PE Ratio(Funding) attribute has undisclosed and unknown value so we can remove them if they are less.**

**Checking how many values are presen with value undisclosed**

```python
data[(data['PE Ratio']=='undisclosed')]
```

**No value is undisclosed in PE Ratio.**

**Checking how many values are present with value unknown**

```python
data[(data['PE Ratio']=='unknown')]
```

**No value in unknown in the PE Ratio Feild.**

**The number of non numerical value for `PE Ratio` attribute are there , so we can remove them.**

```python
data[data['PE Ratio'].isnull()]
```

**Drop rows with non numerical funding amount.**

```python
data.drop(data.loc[(data['PE Ratio']=='undisclosed') | (data['PE Ratio']=='Undisclosed') | (data['PE Ratio'].isnull()) | (data['PE Ratio']=='unknown')].index,inplace=True)

data['PE Ratio'].fillna('None',inplace=True)

data.shape
```

# Data Visualization 

```python
company=[]
for Name in data['Name'].unique():
    df=data[data['Name']==Name]    
    sum=np.sum(df['Price'])        
    company.append([Name,sum])

company=pd.DataFrame(company,columns=['Name','Price'])  #converting the list to dataframe and sort them by the fundin amount
company.sort_values(by='Price',ascending=False,inplace=True)
```

**Visualising the most valued companies.**

```python
industry = data['Name'].value_counts().head(10)
plt.figure(figsize=(15,8))
sns.barplot(industry.index, industry.values, alpha=0.9, palette="husl")
plt.xticks(rotation=45)
plt.xlabel('Companies', fontsize=12)
plt.ylabel('freq of companies', fontsize=12)
plt.title("Most valued companies ", fontsize=16)
plt.show()
```

**Famous companies of all**

```python
d=data[data['Name']!='others']['Name'].value_counts().head(6)
explode = (0.1, 0, 0, 0,0,0)
fig1,ax1=plt.subplots(figsize=(20,10))

ax1.pie(d.values,explode=explode, labels=d.index,autopct='%1.1f%%', shadow=True, startangle=140)
ax1.axis('equal')
plt.title("Famous Companies",fontsize=30)
plt.show()
```
