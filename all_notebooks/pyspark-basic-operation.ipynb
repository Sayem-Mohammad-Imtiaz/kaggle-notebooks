# Pyspark Handling Missing Values
*  Dropping Columns
*  Dropping Rows
* Various Parameter in Dropping Functionality 
* Handling missing value by mean

```python
# importing pandas as pd
import pandas as pd
  
# importing numpy as np
import numpy as np
  
# dictionary of lists
dict = {'First Score':[100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score':[np.nan, 40, 80, np.nan]}
  
# creating a dataframe from list
df = pd.DataFrame(dict)
```

```python
# Converting Pandas datafrmae to spark
df = spark.createDataFrame(df)
```

```python
df.show(5)
```

```python
df.na.drop().show()
```

```python
# any = How 
df.show(5)
```

```python
# any one na value in that column then it will drop
df.na.drop(how="any").show()
```

```python
# Threshold is 2 any row with maximum two null values will get deleted 
df.na.drop(how="any", thresh=2).show()
```

```python
#Subset ie only from specfic column in this we have selected as the column as  Third Score
df.na.drop(how="any",subset=["Third Score"]).show()
```

```python
# Filling the missing values
df.na.fill("Missing Values").show()
```

# Filter operation

```python
df_pyspark.filter("Inv_Amt<4").show()
```

```python
df_pyspark.filter("Inv_Amt<4").select(["Inv_Id","Vendor_Code"]).show()
```

```python
#Another way to write the filter
df_pyspark.filter(df_pyspark["Inv_Amt"]<4).show()
```

```python
#Another way to write the filter with not condition (inverse operation)
df_pyspark.filter(~(df_pyspark["Inv_Amt"]<98)).show()
```

```python
#Multiple condition
df_pyspark.filter((df_pyspark["Inv_Amt"]<4) & (df_pyspark["Inv_Id"]<16000)).show()
```

# Pyspark GroupBy And Aggregate Function 

```python
## Groupby sum of invoice amounta
df_pyspark.groupBy("GL_Code").sum("Inv_Amt").show()
```

```python
## Groupby mean value
df_pyspark.groupBy("GL_Code").mean("Inv_Amt").show()
```

```python
## Groupby max value
df_pyspark.groupBy("GL_Code").max("Inv_Amt").show()
```

```python
## Groupby count
df_pyspark.groupBy("GL_Code").count().show()
```

```python
# aggregate sum of invoice amount
df_pyspark.agg({"Inv_Amt":"sum"}).show()
```
