```python
#Multiple Linear Regression

#import dataset
dataset = pd.read_csv('/kaggle/input/50-startups/50_Startups.csv')
dataset
```

```python
#Create matrix of independent variables
X = dataset.iloc[:,:-1].values

#Create array of dependent variable
y = dataset.iloc[:,4]

#Encoding categorical independent variable(s)
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(
    [('oh_enc',OneHotEncoder(sparse=False),[3]),],
     remainder='passthrough'
     )
X = ct.fit_transform(X)

#Avoid dummy variable trap
X = X[:, 1:]

#Splitting into test set and training set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
#Fit Multiple Linear Regression Model to the Training Set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Predicting the Test Set results
y_pred = regressor.predict(X_test)
```

```python
#Building the optimal model using Backward Elimination
import statsmodels.api as sm
#create a new column of ones as required by 'statsmodels' library
X = np.append(arr=np.ones((50,1)).astype(int), values = X , axis=1)
#X_opt will contain all variables at first, and then remove through backward elimination
X_opt = X[:,[0,1,2,3,4,5]]
#constant, FL, NY, RnD, Admin, MktS
X_opt = X_opt.astype(np.float64)
regressor_OLS = sm.OLS(y,X_opt).fit()
regressor_OLS.summary()
```

```python
#variable x2: New York has the highest p-value so we will proceed with removing it
#from x_opt
X_opt = X[:,[0,1,3,4,5]]
X_opt = X_opt.astype(np.float64)
regressor_OLS = sm.OLS(y,X_opt).fit()
regressor_OLS.summary()
```

```python
#variable x1:Florida has the highest p-value so we will proceed with removing it
#from x_opt
X_opt = X[:,[0,3,4,5]]
X_opt = X_opt.astype(np.float64)
regressor_OLS = sm.OLS(y,X_opt).fit()
regressor_OLS.summary()
```

```python
#variable x2:Admin has the highest p-value so we will proceed with removing it
#from x_opt
X_opt = X[:,[0,3,5]] 
X_opt = X_opt.astype(np.float64)
regressor_OLS = sm.OLS(y,X_opt).fit()
regressor_OLS.summary()
```

```python
#variable x2:MktS has the highest p-value so we will proceed with removing it
#from x_opt
X_opt = X[:,[0,3]]
X_opt = X_opt.astype(np.float64)
regressor_OLS = sm.OLS(y,X_opt).fit()
regressor_OLS.summary()
```

```python
#Removing MktS actually lowered our adj R^2 so we should add it back in
#The optimal model is as follows:
X_opt = X[:,[0,3,5]] #const+RnD+MktS
X_opt = X_opt.astype(np.float64)
regressor_OLS = sm.OLS(y,X_opt).fit()
regressor_OLS.summary()
```
