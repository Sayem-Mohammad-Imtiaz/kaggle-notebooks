```python
fig_opc = pd.DataFrame(df["aeronave_operador_categoria"].value_counts()).reset_index()
fig_opc = fig_opc.rename(columns={'index':'Categorias', 'aeronave_operador_categoria':'Quantidade'})
fig_opc.plot.bar(x='Categorias', y='Quantidade', label='Categorias de Operador de Aeronave', figsize=(20,5))
```

```python
import seaborn as sns

df.describe()

df_fat = df[df['total_fatalidades'] > 0]

dfG = df_fat.groupby(['aeronave_assentos'],as_index = False).sum().sort_values(by='total_fatalidades', 
                                                                         ascending = False)

fig = plt.figure()
ax = fig.add_axes([0,0,2.2,0.9]) 
ax.grid(zorder=0)
ax.set_ylabel('Número de Assentos')
ax.set_xlabel('Fatalidades')
ax.bar(dfG.aeronave_assentos,dfG.total_fatalidades, color=sns.color_palette("rocket"), align='center',zorder=3)
plt.title('Fatalidade por número de Assentos')
plt.show()
```

```python
df_fat = df[df['total_fatalidades'] > 0]
x1 = df['total_fatalidades']
plt.title('Distribuição do total de fatalidades')
plt.boxplot ([x1])
plt.show()
```

```python
df_ano = df_fat[df_fat['aeronave_ano_fabricacao'] > 1950]
df_ano = df_ano.groupby(['aeronave_ano_fabricacao'],as_index = False).sum().sort_values(by='total_fatalidades', 
                                                                         ascending = False)
fig = plt.figure()
ax = fig.add_axes([0,0,2,1])
ax.grid(zorder=0)
ax.set_ylabel('Número de Assentos')
ax.set_xlabel('Fatalidades')
ax.barh(df_ano.aeronave_ano_fabricacao,df_ano.total_fatalidades, align='center')
plt.title('Fatalidade por Ano de fábricação da Aeronave')
plt.show()
```

```python
pais_eua_cessna = df.loc[(df['aeronave_pais_fabricante'] == 'ESTADOS UNIDOS') & (df['aeronave_fabricante'] == 'CESSNA AIRCRAFT') & (df['aeronave_ano_fabricacao'] > 0)]
pais_eua_cessna = pais_eua_cessna[['aeronave_ano_fabricacao','total_fatalidades']]
pais_eua_cessna = pais_eua_cessna.groupby(['aeronave_ano_fabricacao'],as_index = True).sum()#.sort_values(by='total_fatalidades', ascending = False)
pais_eua_cessna
ax = pais_eua_cessna.plot.line(title='Qntd de Mortes por Ano de Fabricação da Fabricante Cessna nos Estados Unidos',subplots=False, figsize=(20,5))
```
