# Let's find that lucky book!!

We will be using **pivot table** for the purpose. It takes the labels in **index** keyword and groups automatically the overlapps, with respect to **values**. We will be using **sum** as aggregate function to determine the book with hishest text review count. Lastly, Let's just sort the list in descending order for better readibilty and print only Top 10. 

```python
df1 = pd.pivot_table(df,index = 'title',values = 'text_reviews_count',aggfunc='sum').sort_values('text_reviews_count',ascending = False)
df1[:10]
```

As you can see, many titles have unwanted characters, we can very easily remove those.
**Approach**: Let's take the title with highest number of text reviews and search it in main dataframe. After that, extract the title, remove unwanted characters and take first string element from array generated. 

```python
Book_max_text_review = df[df['text_reviews_count'] == df1.text_reviews_count.max()]['title'].unique()[0].split('(')[0]
```

Great! Here is the answer, then.

```python
print(Book_max_text_review)
```

It's actually good to see books like **Twilight** and **The Alchemist** in Top-10.
