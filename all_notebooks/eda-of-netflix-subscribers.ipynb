```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
NetflixSubscribersbyCountryfrom2018toQ2_2020_filepath = "../input/netflix2020/NetflixSubscribersbyCountryfrom2018toQ2_2020.csv"

# Read the file into a variable flight_data
Subscribers_data = pd.read_csv(NetflixSubscribersbyCountryfrom2018toQ2_2020_filepath, index_col="Area")
```

```python
# Print the data
Subscribers_data
```

## Pie charts showing the distribution of subscribers across different markets

```python
# Extract Column Names as List in Pandas Dataframe
Subscribers_data_columns =list(Subscribers_data.index) 
my_labels = Subscribers_data_columns
my_labels
```

```python
fig = plt.figure(figsize=(18,10))
fig.suptitle("PIE CHARTS SHOWING DISTRIBUTION OF SUBSCRIBERS IN Q1-2020 AND Q2-2020", fontname='Comic Sans MS' ,fontsize=22)


# Plotting the Q1-2020 Pie Chart
ax1 = plt.subplot2grid((2,2),(0,0))
plt.pie(Subscribers_data['Q1 - 2020'] ,labels=my_labels,autopct='%1.1f%%', textprops={'fontsize': 13.5})
plt.title('Distribution of subscribers as of Q1-2020', fontname='Comic Sans MS', fontsize=18)
plt.axis('equal')



# Plotting the  Q2-2020 Pie Chart
ax1 = plt.subplot2grid((2,2),(0,1))
plt.pie(Subscribers_data['Q2 - 2020'] ,labels=my_labels,autopct='%1.1f%%', textprops={'fontsize': 13.5})
plt.title('Distribution of subscribers as of Q2-2020', fontname='Comic Sans MS', fontsize=18)
plt.axis('equal')
```
