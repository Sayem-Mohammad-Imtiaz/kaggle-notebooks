```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(data.Review,data.Rating,test_size=0.2,random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
```

```python
X=cv.fit_transform(X_train.values)
```

```python
cv.get_feature_names()
```

```python
print(X.toarray())
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model=MultinomialNB()
```

```python
model.fit(X,Y_train)
```

```python
test_review=[
    'This is the best hotel to stay. It offres everything you want.',
    'This is worst hotel'
] #This is simple review only for test purpose
```

```python
test_review_cv=cv.transform(test_review)
```

```python
test_review_cv.toarray()
```

```python
model.predict(test_review_cv)
```

```python
#Model does preety well in this small reviews
```

```python
test_count=cv.transform(X_test.values)
```

```python
model.score(test_count,Y_test)
```

```python
#But it does not perform well for test data. It is good for beginners to practice
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg_model=LogisticRegression()
```

```python
lg_model.fit(X,Y_train)
```

```python
lg_model.score(test_count,Y_test)
```

```python
#Logistic Regression performs slightly better than Naive Bayes
```

```python
lg_model.predict(test_review_cv)
```

```python
#If you don't want to perform count vectorizer everytime you need to predict, There is a method called Pipeline.
# It helps in performing series of operation at single time
```

```python
from sklearn.pipeline import Pipeline
```

```python
clf=Pipeline([('vectorizer',CountVectorizer()),('lg',LogisticRegression())])
```

```python
clf.fit(X_train,Y_train)
```

```python
clf.score(X_test,Y_test)
```
