```python
dataframe = pd.read_csv('/kaggle/input/wisconsin-breast-cancer-cytology-features/wisconsin_breast_cancer.csv')
dataframe=dataframe.fillna(0)
dataframe.info()
```

```python
X= dataframe.iloc[:,1:-1]
X
y= dataframe.iloc[:,-1]
y
```

X and y value

```python
from sklearn.linear_model import SGDClassifier
sgd_clf =SGDClassifier(random_state=1506360)
```

SGDClassifier

```python
from sklearn.model_selection import cross_val_predict
SGD_clf = cross_val_predict(sgd_clf, X, y, cv =5, method="decision_function")
```

cross_val_predict 

```python
from sklearn.ensemble import RandomForestClassifier
forest_clf = RandomForestClassifier(n_estimators= 30, random_state=1506360)
```

Random forest classifier with 30 tree

```python
y_probas_forest = cross_val_predict(forest_clf, X, y, cv=5, method= "predict_proba")
Frost= y_probas_forest[:,-1]
```

```python
from sklearn.naive_bayes import GaussianNB
NB = GaussianNB()
```

gaussianNB

```python
DNB = cross_val_predict(NB, X, y, cv=5, method= "predict_proba")
GNB= DNB[:,-1]
```

```python
from sklearn.metrics import roc_curve
fpr_forest, tpr_forest, thresholds_forest = roc_curve(y,Frost)
```

```python
fpr_sgd, tpr_sgd, thresholds_sgd = roc_curve(y,SGD_clf )
```

```python
fpr_gauss, tpr_gauss, thresholds_gauss = roc_curve(y,GNB )
```

```python
import matplotlib.pyplot as plt
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr,tpr, linewidth=2, label=label)
    plt.plot([0,1],[0,1],'k--')
```

```python
from sklearn.metrics import roc_auc_score
plot_roc_curve(fpr_gauss, tpr_gauss,"GaussClassfier "+str(roc_auc_score(y,SGD_clf)))
plot_roc_curve(fpr_sgd, tpr_sgd, "SGD_Classfier "+str(roc_auc_score(y,Frost)))
plot_roc_curve(fpr_forest, tpr_forest, "Random_Classfier "+str(roc_auc_score(y,GNB)))

plt.legend(loc ="lower right")
plt.xlabel("False postive")
plt.ylabel("True postive")
```
