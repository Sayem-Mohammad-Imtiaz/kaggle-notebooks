```python

import numpy as np
import matplotlib.pyplot as plt1
import matplotlib.pyplot as plt2
import pandas as pd

dataset1 = pd.read_csv("../input/cardiodata/Initial_Train.csv")

set_1=dataset1.head(28000)
set_2=dataset1.tail(28000)

x1 = set_1.iloc[:,[1,4,3,2,5,6,11]].values
x2 = set_2.iloc[:,[1,4,3,2,5,6,11]].values

from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(affinity='euclidean', compute_full_tree='auto',
                                            connectivity=None, distance_threshold=None,
                                            linkage='ward', memory=None, n_clusters=4)

cl1=cluster.fit_predict(x1)
set_1['Cluster']=cl1
set_1.to_csv('Train_hr_set_1.CSV')

cl2=cluster.fit_predict(x2)
set_2['Cluster']=cl2
set_2.to_csv('Train_hr_set_2.CSV')

plt1.scatter(x1[cl1==0,0],x1[cl1==0,1],s=50,c='red',label='cluster 0')
plt1.scatter(x1[cl1==1,0],x1[cl1==1,1],s=50,c='green',label='cluster 1')
plt1.scatter(x1[cl1==2,0],x1[cl1==2,1],s=50,c='blue',label='cluster 2')
plt1.scatter(x1[cl1==3,0],x1[cl1==3,1],s=50,c='yellow',label='cluster 3')
plt1.title('Cluster of Patient')
plt1.legend()
plt1.show()

plt2.scatter(x2[cl2==0,0],x2[cl2==0,1],s=50,c='red',label='cluster 0')
plt2.scatter(x2[cl2==1,0],x2[cl2==1,1],s=50,c='green',label='cluster 1')
plt2.scatter(x2[cl2==2,0],x2[cl2==2,1],s=50,c='blue',label='cluster 2')
plt2.scatter(x2[cl2==3,0],x2[cl2==3,1],s=50,c='yellow',label='cluster 3')
plt2.title('Cluster of Patient')
plt2.legend()
plt2.show()

```
