```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.displot(happy['pop2021'],color='purple',bins=20,kde=True)
```

### the curve is positively skewed.
### the number of countries with least population are high in numbers.

```python
sns.displot(happy['happinessRank'],color='green',bins=30,kde=True)
```

```python
plt.scatter(x='pop2021',y='score',data=happy)
```

```python
sns.regplot(x='pop2021',y='score',data=happy)
```

### from the graph above there is a negative coorelation between the population and score
### As population increases the happiness score decreases

```python
sns.pairplot(happy)
```

### So, in the graph above, we can see the relationships between each of the variables with the other and thus infer which variables are most correlated.
