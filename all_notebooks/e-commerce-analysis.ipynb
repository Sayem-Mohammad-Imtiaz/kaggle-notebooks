```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
data= pd.read_csv("../input/customer-analytics/Train.csv")
data.head()
```

```python
data.drop('ID', axis= 1)
```

```python
data["customer_score"]=np.nan
data.head()
```

```python
Rating = data.query('Customer_rating>=4')
Rating
```

```python
data.rename(columns={"Reached.on.Time_Y.N":"Reached_on_time"}, inplace =True)
```

```python
data ['Reached_on_time'].value_counts(normalize= True)* 100
```

```python
totalcost= data['Cost_of_the_Product'].sum()
print(totalcost)
```

```python
order= data['Cost_of_the_Product'].count()
print(order)
```

```python
average= totalcost/order
print(average)
```

```python
x=[]
for value in data['Prior_purchases']:
    x.append(average*value)
data['customerscore']=x
data.head()
```

```python
data['customerscore'].median()
```

```python
data['Cost_of_the_Product'].max()
```

```python
median=data['Cost_of_the_Product'].median()
print(median)
```

```python

plt.pie(data.Mode_of_Shipment.value_counts(),explode=[.2,.1,.1],startangle=90,autopct='%.2f%%',labels=['ship','flight','Road'],radius=10,)
plt.axis('equal')
plt.title('Mode of shipment',fontdict={'fontsize':20,'fontweight':'bold'})
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.countplot(data.Mode_of_Shipment,palette='Dark2_r')
plt.show()
```

```python
plt.pie(data.Product_importance.value_counts(),explode=[.5,.1,.4],startangle=90,autopct='%.2f%%',labels=['low','medium','High'],radius=10,colors=['#ff00ff','#00ff00','yellow'])
plt.axis('equal')
plt.title('Importance of product',fontdict={'fontsize':20,'fontweight':'bold'})
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.countplot(data.Product_importance,palette='Dark2')
plt.show()
```

```python
plt.pie(data.Gender.value_counts(),explode=[.1,.3],startangle=90,autopct='%.2f%%',labels=['female','male'],radius=10,colors=['red','yellow'])
plt.axis('equal')
plt.title('Gender',fontdict={'fontsize':22,'fontweight':'bold'})
plt.show()
```
