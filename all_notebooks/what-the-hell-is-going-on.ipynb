Let's sort by IMDB Rating.

Let's explore the idea that we can find some good information on the assumption that the best movies have something that we can learn from. 

First. we can use the **sort_values()** function found in Pandas. 

This can pass in a bunch of variables to making sorting easier, but we'll get to that in a second.

If you want to know more about the sort_values function: 

http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.dropna.html

```python
df.sort_values('imdb_score')
```

**Sweet! We've sorted  the Data set by IMDB Rating.**


But wait a minute: here's the problem. It's sorting lowest to highest. That's not good. 

Remember how I talked about the variables you can add for **sort_values()** ? 

**sort_values()** can actually pass in a variable called **ascending = True/False**

REMEMBER TO CAPITALIZE THE FIRST LETTER OF True of False otherwise you run into a problem.

If we said sort_values('imdb_score', ascending=True) -> It will sort from 1.0 - 10.0  or Lowest to Highest.

What we want to do is sort Highest to Lowest, or in DESCENDING order. Therefore, we want to say:

df.sort_values('imdb_score', ascending=False)

```python
df.sort_values('imdb_score', ascending=False)
```

We're making progress, but still a long way to accomplish what we want to see.
