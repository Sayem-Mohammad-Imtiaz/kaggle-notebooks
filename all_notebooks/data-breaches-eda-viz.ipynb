```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly.subplots as sp
```

```python
breaches = pd.read_csv("../input/list-of-top-data-breaches-2004-2021/DataBreaches(2004-2021).csv")
```

```python
breaches
```

```python
breaches.Method.unique()
```

**Replacing Methods that may be equivalent or written in different ways**

```python
breaches["Method"].replace(
    {"inside job, hacked": "inside job",
     'poor security/inside job': "inside job",
     "data exposed by misconfiguration":"misconfiguration",
     "misconfiguration/poor security":"misconfiguration",
     "poor security / hacked": "poor security",
     "Poor security":"poor security",
     "accidentally uploaded":"accidentally published",
     "accidentally exposed":"accidentally published",
    }, inplace=True)
```

### **Yearly Size and Method of Breaches**

**By The Size of Data**

```python
fig = px.histogram(breaches, x="Year", y="Records", color="Method", labels={'x':'Year', 'y':'Records'}).update_xaxes(categoryorder='total descending')
fig.show()
```

**By Number of Breaches**

```python
fig = px.histogram(breaches, x="Year", y="Records", histfunc="count", color="Method", labels={'x':'Year', 'y':'Records'}).update_xaxes(categoryorder='total descending')
fig.show()
```

* **The biggest data breaches in terms of Records lost occured in the year 2019 and 2013 while most of them occured in 2011 and 2019**

### **Affected Sectors**

```python
fig = px.pie(breaches, values ="Records", names ="Organization type", title='Breaches by Sector')
fig.update_traces(textposition='inside')
fig.update_layout(uniformtext_minsize=10, uniformtext_mode='hide')
fig.show()
```

```python
fig =px.sunburst(breaches, path=["Organization type","Method"], values='Records', width = 650, height = 650, title = "Method of Breaches for each Sector")
fig.show()
```

* **We see that web apps followed by social networks, tech companies and financial institutions are the most targeted by data breaches**
* **We also noticed that web apps are the most susceptible to hacking while others sectors suffer from data breaches because of poor security**

### **Most Affected Companies**

```python
fig =px.sunburst(breaches, path=["Organization type","Entity"], values='Records', width = 700, height = 700, title = "Affected Company for each Sector")
fig.show()
```
