# Problem

    You own an ice cream business and you would like to create a model that could predict the daily revenue in dollars based on the outside air temperature (degC). You decided to build a simple Artificial Neural Network to solve this problem.

    Data set:
        Input (X): Outside Air Temperature
        Output (Y): Overall daily revenue generated in dollars.

# Import Libraries

```python
import pandas as pd					# data manipulation using dataframes
import numpy as np					# data statistical analysis

import seaborn as sns				# Statistical data visualization
import matplotlib.pyplot as plt		# data visualisation
%matplotlib inline
```

```python
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.optimizers import Adam
```

# Importing Dataset

```python
raw_df = pd.read_csv("../input/ice-cream-revenue/IceCreamData.csv")
```

```python
df = raw_df.copy()
```

# Data Preprocessing

```python
### Define X&Y ###

X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
### Splitting Dataset ###

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y , test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

# Model Building

```python
model = Sequential()
model.add(Input (shape = (1)))
model.add(Dense (16))
model.add(Dense (1))
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(0.01), loss='mse', metrics = ['mae', 'mse'])
```

# Model Training

```python
%%time
history = model.fit(X, y,
                    epochs = 1000,
                    verbose = 2,
                    validation_split = 0.2 )
```

```python
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
```

```python
model.get_weights()
```

```python
history.history.keys()
```

# Model Prediction

```python
y_pred = model.predict(X_test)
```

```python
test_df = pd.DataFrame(y_pred, columns = ['pred'])
test_df['actual'] = y_test
test_df.head(10)
```

# Model Visualization

```python
# Visualization training results

plt.scatter(X_train, y_train, color = 'blue')
plt.plot(X_train, model.predict(X_train), color = 'red')
plt.xlabel('Revenue')
plt.ylabel('Temperature')
plt.title('Revenue Vs Temperature')
plt.grid(True)
```

```python
# Visualization test results

plt.scatter(X_test, y_test, color = 'blue')
plt.plot(X_train, model.predict(X_train), color = 'red')
plt.xlabel('Revenue')
plt.ylabel('Temperature')
plt.title('Revenue Vs Temperature')
plt.grid(True)
```

# Metrices

```python
### RMSE ###

from sklearn import metrics
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
### R-squared ###

from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
