# **HABERMAN ANALYSIS**

**About the dataset:**
The dataset consists of case study conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.
It contains:
  * 305 datapoints
  * 4 features including age,year of operation,positive auxilary and living status after 5 years

# **OBJECTIVE**
To predict the survival of a patient after breast cancer surgery.

```python
#Renaming the attributes for better undertanding
df = df.rename(columns = {"30" : "age", "64" : "operation_year", "1" : "axillary_lymph_node", "1.1" : "survival_status"})
df.head()
```

```python
# The values are mapped as follows. 1 :-> 'True' 2 :-> 'False'
df['survival_status'] = df['survival_status'].map({1:True, 2:False})
df.head()
```

**Univariant Analysis**

Since by using a 2d-Scatter-Plot, it is difficult to distinguishing the data points. Adding a third parameter to colorixe the data will improve the readability. Hence using seaborn:

```python
sns.set_style("whitegrid");
sns.FacetGrid(df, hue="survival_status", size=4).map(plt.scatter, "age", "axillary_lymph_node").add_legend();
plt.show();
```

Scince the survival_status based on Age and number of axillary_lymph_node is mixed together without any significant seperation,a pair plot can give us the insight into all relations.

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(df, hue="survival_status", size=3);
plt.show()
```

To see the relation between the attributes, we can go with the correlation matrix.

```python
corr = df.corr(method = 'spearman')
sns.heatmap(corr,xticklabels=corr.columns.values,yticklabels=corr.columns.values)
```

From the above correlation matrix we can see that survival_status is more correlated with axillary_lymph_node. The negative correlation implies that the relation is inversely proportional. ( More lymp nodes = less chance of survival)

**Conclusion**
    * The dataset contains very less data points and is highly imbalanced.
    * The given dataset is not linearly seprable form each class. There are too much overlapping in the data-points and hence it is very diffucult to classify. 
    * To model this data non-linear models have to be used.
**The survival status mainly depends on auxilary nodes as the age and operation's year factor seems to overlap a lot and becoming insignificant in analysis.**
