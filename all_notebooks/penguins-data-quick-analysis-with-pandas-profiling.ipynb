# importing packages

```python
import pandas as pd
from pandas_profiling import ProfileReport
```

# loading data and forming dataframe

```python
df = pd.read_csv("../input/palmer-penguins/penguins.csv")
df.head(2)
```

# forming ProfileReport

```python
profile = ProfileReport(df, title='Penguins Data Report')
```

```python
profile.to_widgets()
```

# using ProfileReport as html-json-file

```python
profile.to_file("penguins_report.html")
```

```python
# As a string
json_data = profile.to_json()
```

```python
# As a file
profile.to_file("penguins_report.json")
```
