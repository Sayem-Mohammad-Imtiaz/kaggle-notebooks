```python
dfn=df.drop('region',axis=1)
```

# Label Encoding

```python
dfn['sex'].replace('female',1,inplace=True)
dfn['sex'].replace('male',0,inplace=True)
dfn['smoker'].replace('yes',1,inplace=True)
dfn['smoker'].replace('no',0,inplace=True)
```

```python
dfn
```

```python
dfn.head()
```

# All features selected

```python
from sklearn.model_selection import train_test_split
X=dfn.drop('charges',axis=1)
y=dfn['charges']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,)
```

```python
import statsmodels.api as sm
XX = sm.add_constant(X_train)
print(XX)
```

```python
#checking p values i.e the significance of the feature
est = sm.OLS(y_train, XX)
est2 = est.fit()
print(est2.summary())
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
pred=lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,pred)
```
