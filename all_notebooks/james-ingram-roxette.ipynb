```python
plt.figure(figsize=(10,10))
plt.title('James Ingram & Roxette')
ax=sns.heatmap(df.corr(),
               linewidth=2.6,
               annot=True,
               center=1)
```

```python
#codes from Mikey_Mtk @motokinakamura https://www.kaggle.com/motokinakamura/treemap-with-plotly
#make a df it's grouped by "Genre"
gb_genre =df.groupby("Genre").sum()

gb_genre.head()
```

```python
# codes from Mikey_Mtk @motokinakamura https://www.kaggle.com/motokinakamura/treemap-with-plotly
genre = list(gb_genre.index)
score = list(gb_genre.Popularity)

print(genre)
print(score)
```

```python
#codes from Mikey_Mtk @motokinakamura https://www.kaggle.com/motokinakamura/treemap-with-plotly
#first treemap
test_tree = go.Figure(go.Treemap(
    labels =  genre,
    parents=[""]*len(genre),
    values =  score,
    textinfo = "label+value"
))

test_tree.show()
```

```python
#codes from Mikey_Mtk @motokinakamura https://www.kaggle.com/motokinakamura/treemap-with-plotly
#second treemap
test_tree_blue = go.Figure(go.Treemap(
    labels =  genre,
    parents=[""]*len(genre),
    values =  score,
    textinfo = "label+value",
    marker_colorscale = 'Cividis'
))

test_tree_blue.show()
```

```python
df['Genre'].value_counts().plot.bar()
plt.title('James Ingram & Roxette')
plt.ylabel('frequency')
plt.xlabel('Genre')
plt.show
```

```python
df = df.rename(columns={'Track.Name':'track', 'Artist.Name': 'artist', 'Beats.Per.Minute': 'beats'})
```

Codes below from https://www.kaggle.com/andresionek/how-to-create-award-winning-data-visualizations

```python
import plotly.express as px

# Grouping it by Genre and track
plot_data = df.groupby(['Genre', 'track'], as_index=False).artist.sum()

fig = px.bar(plot_data, x='Genre', y='artist', color='track')
fig.update_layout(
    title_text='James Ingram and Roxette',
    height=500, width=1000)
fig.show()
```

```python
import plotly.express as px

# Grouping it by Genre and track
plot_data = df.groupby(['Genre', 'track'], as_index=False).artist.sum()

fig = px.line_polar(plot_data, theta='Genre', r='artist', color='track')
fig.update_layout(
    title_text='James Ingram and Roxette',
    height=500, width=1000)
fig.show()
```
