```python
import numpy as np
import pandas as pd
import random
import category_encoders as ce

from sklearn.metrics import classification_report, log_loss, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import ShuffleSplit

from xgboost import XGBClassifier
```

```python
data = pd.read_csv("../input/heart-attack-analysis-prediction-dataset/heart.csv")
data[0:5].T
```

```python
n=len(data)
print(n)
```

```python
N=[]
for i in range(n):
    N+=[i]
random.shuffle(N)
```

```python
output=data['output']
data2=data.drop('output',axis=1)
```

```python
output.value_counts()
```

```python
output=np.array(output)
data2=np.array(data2)
```

```python
X=data2[N[0:(n//10)*8]]
y=output[N[0:(n//10)*8]]

test=data2[N[(n//10)*8:]]
test_output=output[N[(n//10)*8:]]
```

```python
clf = XGBClassifier(max_depth=3, n_estimators=1000, learning_rate=0.01)
```

```python
ss = ShuffleSplit(n_splits=5,train_size=0.8,test_size=0.2,random_state=0) 

for train_index, test_index in ss.split(X): 
    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = y[train_index], y[test_index]
    clf.fit(X_train, Y_train) 
    print(clf.score(X_test, Y_test))
```

```python
pred2 = clf.predict(np.array(test))
```

```python
PRED=pred2   
ANS=test_output
```

```python
accuracy=accuracy_score(ANS,PRED)
print(accuracy)
```
