```python
# 3. Sex
plt.figure(figsize=(3,3))
df.Sex.value_counts().plot.pie()
print(df.Sex.value_counts())
```

```python
# transform data from str to float
d = {"M": 0, "F": 1}
df['Sex'] = df.Sex.apply(d.get).astype(float)
```

```python
# 4. Race
plt.figure(figsize=(3,4))
df.Race.value_counts().plot(kind='bar', rot=0)
plt.ylabel('crime counts')
print(df.Race.value_counts())
```

We'll consider later whether we have to do one-hot-encoding.

```python
# 5. ArrestDate
print('the data type of this feature is', type(df.loc[0, 'ArrestDate']))
# transform to datetime
df['ArrestDate'] = pd.to_datetime(df['ArrestDate'], format= '%m/%d/%Y')
print('the new data type of this feature is', type(df.loc[0, 'ArrestDate']))
```

```python
# 6. ArrestTime
print('the data type of this feature is', type(df.loc[0, 'ArrestTime']))
# transform to datetime
df['ArrestTime'] = df['ArrestTime'].str.replace(r'.',':')
df['ArrestTime'] = pd.to_datetime(df['ArrestTime'], format='%H:%M')
# actually hour is enough for our analysis
df['ArrestHour'] = df['ArrestTime'].apply(lambda x: x.hour)
print('the new data type of this feature is', type(df.loc[0, 'ArrestHour']))
```

```python
df.ArrestHour.value_counts().sort_index(ascending=True).plot(kind='bar', rot=0);
plt.xlabel('hours');
plt.ylabel('crime counts');
```

```python
# 7. ArrestLocation
print('number of NaN in this column: ', df.ArrestLocation.isnull().sum())
# more than one third of total entries, it might be better not to use this feature
```

```python
# 8. IncidentOffense
df.loc[df.IncidentOffense=='UNKNOWN OFFENSE', 'IncidentOffense'] = 'Unknown Offense'
OffenseType = df.IncidentOffense.value_counts()
```

```python
print('number of incident types: ', OffenseType.shape[0])
print('number of unkown offense: ', OffenseType['Unknown Offense'])
print('number of incidents only existing once: ', OffenseType[OffenseType==1].shape[0])
print('number of incidents existing less than 10 times: ', OffenseType[OffenseType<10].shape[0])
# we'll decide later how to do with this feature
```

```python
# 15. Location1
df['Latitude'] = df['Location 1']
df['Longtitude'] = df['Location 1']

hi = df.loc[df['Location 1'].notnull(), 'Location 1']
hello1 = pd.Series([x[1:-1].split(',')[0] for x in  hi], index=hi.index)
hello2 = pd.Series([x[1:-1].split(',')[1] for x in  hi], index=hi.index)

df.loc[df['Location 1'].notnull(), 'Latitude'] = hello1
df.loc[df['Location 1'].notnull(), 'Longtitude'] = hello2

df.Latitude = df.Latitude.astype(float)
df.Longtitude = df.Longtitude.astype(float)

print('number of NaN: ', df.Latitude.isnull().sum())
```

We can see that numbers of NaN of all location features are almost the same, so...
