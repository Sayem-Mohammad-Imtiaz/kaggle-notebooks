```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
infile = '../input/Health_AnimalBites.csv'
```

```python
df = pd.read_csv(infile)
df.head()
```

```python
d2 = pd.concat([df[df.SpeciesIDDesc == 'CAT'], df[df.SpeciesIDDesc == 'DOG']]).reset_index(drop=True)
```

```python
d2.head()
```

This table shows the percentage of reports that belong to cats and dogs:

```python
n = d2["SpeciesIDDesc"].value_counts() / len(df) * 100
n
```

This graph is a visual representation of the table above:

```python
n.plot(kind="bar")
plt.yticks(np.arange(0,101,10))
plt.title("Dog and Cat Bite Reports")
plt.ylabel("Percentage of Bites")
```

Shockingly, 78% of this dataset is comprised of dog bites!  Cat bites make up an insignificant 17% of these reports. It really makes sense though. As we all know, cats are little angels. So I would argue that these data are in agreement that cats are superior animals compared to dogs.
