```python
#We have to use categorical or integer type 
#Lets convert object type data.class to integer 
#Abnormal = 1 , Normal = 0
data["class"] = [1 if i == "Abnormal" else 0 for i in data["class"]]
```

```python
plt.figure(figsize = (6.5,6.5))
sns.countplot(data["class"])
print(data["class"].value_counts())
plt.show()
```

```python
A = data[data["class"] == 1] #abnormal
N = data[data["class"] == 0] #normal
```

```python
#visualization
f,ax = plt.subplots(figsize=(10,6))
sns.set_style("whitegrid")
sns.despine(f, left = True , bottom = True)

sns.scatterplot(A["pelvic_radius"] , A["lumbar_lordosis_angle"] , color = "red" , label = "Anormal" , alpha = .7 )
sns.scatterplot(N["pelvic_radius"] , N["lumbar_lordosis_angle"] , palette ="BuPu" , label = "Normal")
plt.xticks(np.arange(20,200,20))
plt.yticks(np.arange(20,200,20))
plt.show()
```

```python
#dependent and independent variables
y = data[["class"]] 
x = data.drop("class" ,axis = 1)
```

```python
#normalization
#x = (x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data))
#normalization yapınca, accuracy değeri düşüyor. (anlayamadım)
```

```python
#importing libraries
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
#train-test splitting
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.3, random_state = 1)
#random_state = 42 olunca accuracy değeri düşüyor..(yine anlamadım)
```

```python
#knn 
knn = KNeighborsClassifier(n_neighbors=3) # choosing k value = 3 randomly
knn.fit(x_train,y_train)

#accuracy score 
print("Accuracy score with k value of {} : {}".format(3,knn.score(x_test,y_test)))
```

```python
#The best accuracy score with k value by visualization
accuracy_score = []
for i in range(1,50):
    knn_ = KNeighborsClassifier(n_neighbors=i)
    knn_.fit(x_train,y_train)
    accuracy_score.append(knn_.score(x_test,y_test))

#visualization
f,ax = plt.subplots(figsize = (12,6))
sns.despine(f, left = True , bottom = True)
plt.plot(np.arange(1,50),accuracy_score)
plt.xticks(np.arange(1,50,1))
plt.show()
```

```python
#knn 
knn = KNeighborsClassifier(n_neighbors=13) # choosing k value = 13 looking at the graph.
knn.fit(x_train,y_train)

#accuracy score 
print("Accuracy score with k value of {} : {}".format(13,knn.score(x_test,y_test)))
```
