```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv('../input/retail-business-sales-20172019/business.retailsales.csv')

```

```python
#drawing histogram of features
data.hist(bins=30, figsize=(20,20))
```

```python
#drwaing corellation bar
corellations=data.corr()
f,ax=plt.subplots(figsize=(20,20))
sns.heatmap(corellations,annot=True)
```

```python
#obervations 
#we can find a strong negative corelation between discounts and total net sales which means if discounts are reduced total net sales will gradually decrease
#we find a negative corellation between gross sales and rerurns which means we have to increase the gross sales or total net sales to increase the returns 
```

```python
#Finding net quantity demand per product-type
data.groupby('Product Type')['Net Quantity'].sum()
```

```python
##here we can find mean quantity ordered by per customer for each product-type
data.groupby('Product Type')['Net Quantity'].mean()
```

```python
#observations
# 1: we find highest customer demand in Art & Sculpture  and Basket  with 1427 and 1461 net quantity ordered respectively
# 2:we find second highest demand in Jewelry and kitchen with 991 and 809 net quantity ordered repestively 
```

```python
##Finding total net sales per product-type
data.groupby('Product Type')['Total Net Sales'].sum()
```

```python
#obervations
# 1:highest net sales in Rs are seen in Art and Sculpture and Basket
# 2: second highest net sales are seen in Jewelry and Home Decor 
```

```python
##Finding mean of total net sales per product-type to find least expected turnover per product-type
data.groupby('Product Type')['Total Net Sales'].mean()
```

```python
#finding total discount given per product-type
data.groupby('Product Type')['Discounts'].sum()
```

```python
#finding mean discount given per product-type
data.groupby('Product Type')['Discounts'].mean()
```

```python
##Finding total net sales per product-type
data.groupby('Product Type')['Returns'].sum()
```
