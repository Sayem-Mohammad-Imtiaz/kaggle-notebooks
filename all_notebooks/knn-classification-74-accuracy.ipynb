```python
# import libraries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
# import data 
data=pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
```

```python
data.head()
```

```python
data.Outcome.value_counts()
# there are 500 healthy people and 268 sick people 
```

```python
sick=data[data.Outcome==1]
healthy=data[data.Outcome==0]
```

```python
# visualization about healthy and sick people
plt.scatter(sick.Glucose,sick.Pregnancies,color='red',label='sick')
plt.scatter(healthy.Glucose,healthy.Pregnancies,color='green',label='healthy')
plt.xlabel('Glucose')
plt.ylabel('Pregnancies')
plt.legend()
plt.show()
```

```python
# knn with sklearn 
y=data.Outcome.values
x_data=data.drop(['Outcome'],axis=1)
```

```python
# normalization  features
x=(x_data -np.min(x_data))/(np.max(x_data)-np.min(x_data))
```

```python
# train test split 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
## knn model 
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=50) # k degeri 
knn.fit(x_train,y_train)
prediction=knn.predict(x_test)
```

```python
print('{} knn score {}'.format(50,knn.score(x_test,y_test)))
```

```python
# find k value 
knn2=KNeighborsClassifier()
score=[]
for i in range(1,50):
    knn2=KNeighborsClassifier(n_neighbors=i)
    knn2.fit(x_train,y_train)
    knn2.score(x_test,y_test)
    score.append( knn2.score(x_test,y_test))
plt.plot(range(1,50),score)
plt.xlabel(' k values')
plt.ylabel('accuracy')
plt.show()
```
