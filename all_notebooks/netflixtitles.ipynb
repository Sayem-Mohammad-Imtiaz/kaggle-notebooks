```python
# I will select the columns that I will be working with
working_data = netflix_df[['show_id','type','title','date_added','release_year','rating']]
```

```python
# drop na's from dataset
drop_nas_working_df = working_data.dropna().reset_index(drop=True)
```

```python
# import libraries for graphing

import plotly.graph_objs as go
import plotly.offline as py

py.init_notebook_mode(connected=True)
```

```python
# data set name is too long, i will copy dataset to a new variable

graph_df = drop_nas_working_df.copy()
```

```python
# create a pie chart for the rating column 

rating_counts = graph_df['rating'].value_counts(sort = True)

pie = go.Pie(labels = rating_counts.index, values = rating_counts.values)

layout = go.Layout(title  = 'Rating', font = dict(size=10), legend=dict(orientation = 'h'))

fig = go.Figure(data =[pie], layout = layout)

py.iplot(fig)
```

```python
# check for unique categories in type column
graph_df['type'].unique()
```

```python
# get data ready for bar chart by movie and tv show
movie = graph_df[graph_df['type'] == 'Movie']
tvshow = graph_df[graph_df['type'] == 'TV Show']
```

```python
movie.groupby(['rating'], sort = True)['type'].count()
```

```python
tvshow.groupby(['rating'], sort = True)['type'].count()
```

```python
# create bar chart
movie_bar = go.Bar(name = 'Movie Rating',
                  x = movie.groupby(['rating'])['show_id'].count().index,
                  y = movie.groupby(['rating'])['show_id'].count().values)

tvshow_bar = go.Bar(name = 'TV Show Rating',
                   x = tvshow.groupby(['rating'])['show_id'].count().index,
                   y = tvshow.groupby(['rating'])['show_id'].count().values)

layout = go.Layout(title = 'Movies & TV Shows Ratings')

fig = go.Figure( data = [movie_bar, tvshow_bar], layout = layout)

py.iplot(fig)
```
