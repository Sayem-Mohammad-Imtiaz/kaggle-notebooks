# Data Cleaning

```python
df.isnull().sum()
```

```python
#There is only 2 missing values, so we can drop the 2 missing rows from the data set
df.dropna(inplace=True)
```

# Import Libraries for Exploratory Data Analysis

```python
#import Visualizing libraries
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS
```

```python
target_count = df['target'].value_counts()
```

```python
px.bar(x=target_count.index[:15,], y=target_count[:15], text=target_count.index[:15], title='Top 15 Targets')
```

# Word Cloud to view the Targets

```python
wc = WordCloud(width=1400, height =800, stopwords=STOPWORDS,background_color='white').generate(text=" ".join(df['target']))
plt.figure(figsize=(30,10))
plt.imshow(wc)
```

# WordCloud to view the insutling words

```python
wc = WordCloud(width=1400, height =800, stopwords=STOPWORDS,background_color='white').generate(text=" ".join(df['insult']))
plt.figure(figsize=(30,10))
plt.imshow(wc)
```

# Top 15 highly targets

```python
px.pie(target_count[:15], values=target_count[:15], names=target_count.index[:15], title='Trump target')
```
