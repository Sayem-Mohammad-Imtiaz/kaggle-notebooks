```python
data = pd.read_csv("/kaggle/input/imdb-extensive-dataset/IMDb movies.csv")

print(data.columns)
```

```python
data = data[['director', 'actors']]
```

```python
def clean(x):
    answer = []
    for word in x:
        answer.append(word.replace("[","").replace("]","").replace("\'","").replace("\"","").strip("\\"))
    return answer

def cartesian(x):
    answer = []
    for w1 in x[0]:
        for w2 in x[1]:
            answer.append((w1,w2))
    return answer

directors = []
actors = []
data["actors"] = data["actors"].apply(lambda x: str(x).split(", ")).apply(clean)
data["director"] = data["director"].apply(lambda x: str(x).split(", ")).apply(clean)
data["cartesian"] = data[["actors", "director"]].apply(cartesian, axis=1)
data = data.explode('cartesian').reset_index(drop=True)
data["actors"] = data["cartesian"].apply(lambda x: x[0])
data["director"] = data["cartesian"].apply(lambda x: x[1])
del data["cartesian"]
print(data)
    
```

```python
data = data.merge(data, on ="actors")
data
```
