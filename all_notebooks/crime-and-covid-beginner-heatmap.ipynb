```python
# I start by including essential code that's required to get the libraries up and running
# It's not important for you to know what this is..

# ..but is important to know that you need it.

import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

Then I add the following. This gives a name to the csv and then in turn a name to that:

```python
# Path of the file to read
borough_crime_filepath = "../input/mps-borough-level-summary-v02/20212208MPS_borough_level_summary_v01.csv"

# Read the file into a variable flight_data
borough_crime_data = pd.read_csv(borough_crime_filepath, index_col="LookUp_BoroughName")
print("Setup Complete")
```

This is summarized data that I've first taken data from [here](https://www.kaggle.com/benjaminfox/london-crime-records-2019-08-01-to-2021-07-31)then summarised in Sheets. It was then reimported. 

```python
# As this isn't a massive amount of data, we can examine it in it's entirety by 
# simply typing the name

borough_crime_data

# If we wanted the beginning or end we could have used the following:

# borough_crime_data.head()
# borough_crime_data.tail()
```

Okay, so now for our first visual. I decided a heatmap would be nice here partly because I'm not sure what else to use right now for an overview. 

My first question was to see if there was any high level snapshots.

```python
# Heatmap 1 

# We first provide a size descriptor and a title to the map
# These two lines are optional. Experiment.

plt.figure(figsize=(16,16))
plt.title("Heatmap of crime by London Borough, by Month")

# Then the heatmap itself
sns.heatmap(data=borough_crime_data, annot=False)
plt.xlabel("Year / Month")
plt.ylabel("Borough")
```

Okay, success, we are beginning to get somewhere, but what the heck is Westminster doing? 

It seems to be in a world of its own! 

Could this be an outlier?

With its high reporting figures, I decided to remove Westminster Borough with a caveat that it has been omitted. 

It is one thing to simply remove data for no reason, but I simply want to see how this effects a new heatmap.

```python
# Heatmap 2 - A reimport without Westminster Borough - Simply removed from the imported csv file

# Path of the New file
borough_crime_filepath_ww = "../input/mps-borough-summary-without-westminster/20212208MPS_borough_summary_v03_without_Westminster.csv"

# Read the file into a new name called borough_crime_data_ww
borough_crime_data_ww = pd.read_csv(borough_crime_filepath_ww, index_col="LookUp_BoroughName")
print("Setup Complete")

# Size and Title
plt.figure(figsize=(16,16))
plt.title("Heatmap of crime by London Borough without Westminster, by Month")

# Heatmap
sns.heatmap(data=borough_crime_data_ww, annot=False)
plt.xlabel("Year / Month")
plt.ylabel("Borough")
```

Okay, pretty cool. We can begin to see some patterns more clearly. 

There seems to be a two notable drops across the boroughs, one in Mar-Apr 2020 and another in Jan-Feb 2021. This is delineated by the two dark plum columns.

I suggest a possible reason for this. 

On 26 March 2020 the UK lockdown started in relation to COVID ([source](https://www.instituteforgovernment.org.uk/sites/default/files/timeline-lockdown-web.pdf)). Also, the third lockdown happened in January 2021.
