```python
data = pd.read_csv("../input/tesla-stock-price/Tesla.csv - Tesla.csv.csv")
```

```python
data.Open.plot(kind = "line", alpha = .5, color = "purple")
plt.xlabel("dates")
plt.ylabel("open prices")
plt.title("tesla ")
plt.show()
```
