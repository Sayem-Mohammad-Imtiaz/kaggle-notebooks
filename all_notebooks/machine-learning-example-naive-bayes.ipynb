Bu iki feature çıkarıldı ve veriseti bu şekilde güncellendi.

Veri seti içerisinde diagnosis feature da yer alan değerler M ve B şeklinde olduğundan bu değerleri 1 ve 0 a çevirerek object olan veri tipini numerik bir degere donusturuldu. 
M ifadesi tumorun kötü huylu olduğunu B ise iyi huylu olduğunu belirtiyor. Veri setindeki M ler 1 B ler 0 yapıldı. M:Malignant (Kötü huylu tümör) B:benign (İyi huylu tümör)

```python
veri.diagnosis = [1 if each == "M" else 0 for each in veri.diagnosis]
```

y eksenine diagnosis değerleri yüklendi. x eksenine ise diagnosis dışındaki diğer tüm veriler yüklendi

```python
y = veri.diagnosis.values
x_data = veri.drop(["diagnosis"],axis=1)
```

Normalizasyon : veri seti içerisinde bazı feature değerleri çok yüksek olduğundan düşük değerlerdeki feature ların etkinliğini baskılayacağı için oluşturulan model sağlıklı olmayacaktır.
bunu gerçekleştirmek için veriseti normalize edilir.

```python
x = (x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data)).values
```

```python
x.head()
```

Görüldüğü üzere tüm değerler 0 ile 1 aralığına normalize edildi.

Veri seti train ve test olarak ayrıldı. %60 train %20 test ve %20 validate olarak ayrıldı. Bunun için train_test_split fonksiyonu kullanıldı

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
x_train, x_validate, y_train, y_validate = train_test_split(x_train,y_train,test_size=0.25, random_state=42)
```

1. train_test_split fonksiyonu verisetini random olarak train ve test olarak bölmektedir. Yani %60 lik train bölümü verisetinin başından itibaren %60 lık bölüm değildir. Tüm veri içerisinde %60'a tekabül eden miktardır. random_state=42 ifadesi ise bu fonksiyonun her çalıştırıldığında aynı %60 ve %20 lik bölümlerin kullanılmasını sağlıyor. Bu sayede yapılan çalışma farklı kişiler tarafından farklı zamanlarda çalıştırıldığında aynı çıktıları üretiyor. 42 ifadesi ise Bir Otostopçunun Galaksi Rehberi kitabında hayatın anlamı olarak geçen sayı. 42 yerine 20, 19 da kullanılabilir.

SkitLearn kütüphanesinden Naive Bayes Algoritmasını çağırarak x_train  ve y_train değerleri ile algoritma eğitilir.

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(x_train,y_train)
```

```python
y_pred = nb.predict(x_test)
print("Tahmin Edilen Deger: ",y_pred)
```

```python
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
print("Confussion Matrisi: ",confusion_matrix(y_test, y_pred))
print("Dogruluk oranı: ",metrics.accuracy_score(y_test,y_pred))
print("f1 Score(macro): ",f1_score(y_test, y_pred, average='macro'))
print("f1 Score(micro): ",f1_score(y_test, y_pred, average='micro'))
print("f1 Score(Agirliklandirilmis): ",f1_score(y_test, y_pred, average='weighted'))
print("Kesinlik Değeri(macro): ",precision_score(y_test, y_pred, average='macro'))
print("Kesinlik Değeri(micro): ",precision_score(y_test, y_pred, average='micro'))
print("Kesinlik Değeri(Agirliklandirilmis): ",precision_score(y_test, y_pred, average='weighted'))
```
