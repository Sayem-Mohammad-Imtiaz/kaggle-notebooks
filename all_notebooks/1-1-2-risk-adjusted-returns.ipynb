```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

# 1. Read the Dataframe and pre-process

```python
returns=pd.read_csv('../input/edhec-data-for-portfolio-construction-with-python/Portfolios_Formed_on_ME_monthly_EW.csv',
                   header=0,
                   index_col=0,
                   parse_dates=True,
                    na_values=-99.99)
returns.head()
```

3 category by market cap (size of the company):
1. Lo 30 
2. med 40
3. Hi 30

Similar notion for the Qnt and etc 

e.g. Dec i = divide into 10 categories

```python
# only choose 2 interest columns
columns=['Lo 20','Hi 20']
returns=returns[columns]
returns.head()
```

```python
# convert it into real number from %
returns=returns/100
returns.head()
```

# 2. Volatility and annualized_vol

```python
# returns.columns=['SmallCap','LargeCap']
returns.plot.line()
```

```python
returns.std()
```

```python
# annualized_vol from the monthly returns

from math import sqrt

num_of_month = 12
annualized_vol=returns.std()*sqrt(num_of_month)
annualized_vol
```

# 3. Monthly and Annualized Return

```python
# Annualized return 
n_months=returns.shape[0]

annualized_return=(returns+1).prod()**(12/n_months)-1
annualized_return
```

# 4. Return on Risk ration

```python
annualized_return/annualized_vol
```

# 5. Sharpe Ratio

![Sharpe Ratio](https://www.robeco.com/images/c/2/9/c29b4590289388f4c62de5e7bf7ea9db_sharpe-ratio_tcm19-4944_800x0.jpg)

```python
riskfree_rate=0.03 # assume 3%
excess_return=annualized_return-riskfree_rate
sharpe_ratio=excess_return/annualized_vol
sharpe_ratio
```
