```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import dates


# Data
df=pd.read_csv('../input/covid19-cases-deaths-in-icu-malaysia/all_icu_death_data.csv', parse_dates=['date'], dayfirst=True)

# plot size
plt.figure(figsize=(20,10))

# multiple line plots
plt.plot( 'date', 'patient_in_icu', data=df)
plt.plot( 'date', 'patient_on_ventilator_support', data=df)
plt.plot( 'date', 'new_death_cases', data=df)
plt.plot( 'date', 'brought_in_dead', data=df)


# show legend
plt.legend()

# show graph
plt.show()
```
