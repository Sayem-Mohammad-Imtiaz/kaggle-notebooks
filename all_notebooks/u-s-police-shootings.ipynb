```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('../input/us-police-shootings/shootings.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#Convert col to datetime format

df['date'] = pd.to_datetime(df['date'])
df.info()
```

```python
#To drop undesired cols

df.drop(['id', 'name'], axis = 1, inplace = True)
df.head()
```

```python
#To plot pie chart

counts = df['race'].value_counts().sort_index()
print(counts)
# Plot a pie chart
counts.plot(kind='pie', title='Race',figsize=(10,8))

plt.legend()
plt.show()
```

```python
#To plot pie chart

counts = df['manner_of_death'].value_counts().sort_index()
print(counts)
# Plot a pie chart
counts.plot(kind='pie', title='manner_of_death',figsize=(10,8))

plt.legend()
plt.show()
```

```python
#Top 10 arms based on value_counts

plt.figure(figsize = (10,7))
chains = df['armed'].value_counts()[:10] #top 10
sns.barplot(x = chains, y = chains.index)
plt.title("Armed")
plt.xlabel("Count")
```
