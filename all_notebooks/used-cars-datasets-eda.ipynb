## Import 

```python
#import the importance library
#----------------------------

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
#Preproc StandardScaler
from sklearn.preprocessing import StandardScaler
```

## Data EDA 

```python
#read the CSV file 
df_=pd.read_csv(r'../input/used-cars-in-saudi-arabia/used_car_cleaned.csv')
```

```python
df_
```

```python
numerical = pd.DataFrame(df_.select_dtypes(exclude=['object' , 'category']))
numerical
```

```python
ss = StandardScaler()
numeric_col=ss.fit_transform(numerical)

scaled_df2=pd.DataFrame(numeric_col)
scaled_df2.columns=numerical.columns
fig, ax6 = plt.subplots(figsize=(16,8))
sns.boxplot(data=scaled_df2,ax=ax6)
plt.title('Boxplots of numerical features');
```

```python
fig = plt.figure(figsize=(10,5))
sns.countplot(x='car_brand', data=df_)
sns.set_theme(style="whitegrid")
plt.xticks(rotation=70)
plt.title('car_brand');
plt.xlabel('car_brand');
plt.ylabel('Frequency');
```

```python

fig = plt.figure(figsize=(10,5))
sns.countplot(x='car_transmission', data=df_)
sns.set_theme(style="whitegrid")
plt.xticks(rotation=70)
plt.title('car_transmission');
plt.xlabel('car_transmission');
plt.ylabel('Frequency');
```
