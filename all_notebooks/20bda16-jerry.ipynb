```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import RandomizedSearchCV
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor
import time
```

```python
# CatBoostRegressor
tic = time.time()
CGB = CatBoostRegressor(logging_level='Silent')    
CGB.fit(X_train, y_train)
print("time (sec):" + "%6.0f" % (time.time() - tic))
result = mean_squared_error(y_test, CGB.predict(X_test))
```

```python
CGB.fit(X_train,y_train)
test_score=CGB.score(X_test,y_test)
train_score=CGB.score(X_train,y_train)
y_pred_xgb=CGB.predict(X_test)
print("\nThe test score is:" ,test_score)
print("\nThe train score is:" ,train_score)
```

```python
(pd.Series(CGB.feature_importances_, index=X.columns).nlargest(20).plot(kind='barh'))  
plt.show()
```

```python
# XGBRegressor: Hyperparameter tuning with RandomizedSearchCV
tic = time.time()
param_grid ={'learning_rate': [0.02,0.025], 'max_depth': [5,6,7],'n_estimators': [500],'colsample_bytree': [0.9], 'subsample': [0.7], 'tree_method': ["hist"] } 
XGB_random_search = RandomizedSearchCV(XGBRegressor(),param_grid, scoring='neg_mean_squared_error', cv=4,  n_iter=5, random_state=123)
XGB_random_search.fit(X_train,y_train)
print("Best parameters:",XGB_random_search.best_params_)
XGB = XGBRegressor(**XGB_random_search.best_params_)    
XGB.fit(X_train, y_train)
print("time (sec):" + "%6.0f" % (time.time() - tic))
result = mean_squared_error(y_test, XGB.predict(X_test))
print("MSE:" + "%6.2f" % result)
```

```python
XGB.fit(X_train,y_train)
test_score=XGB.score(X_test,y_test)
train_score=XGB.score(X_train,y_train)
y_pred_xgb=XGB.predict(X_test)
print("\nThe test score is:" ,test_score)
print("\nThe train score is:" ,train_score)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred_xgb))
```

```python
tic = time.time()
param_grid ={'learning_rate': [0.02,0.025], 'n_estimators': [500], 'num_leaves': [30,40,50],'feature_fraction': [0.7]} 
LGB_random_search = RandomizedSearchCV(LGBMRegressor(),param_grid, scoring='neg_mean_squared_error', cv=4,  n_iter=5, random_state=123)
LGB_random_search.fit(X_train,y_train)
print("Best parameters:",LGB_random_search.best_params_)
LGB = LGBMRegressor(**LGB_random_search.best_params_)    
LGB.fit(X_train, y_train)
print("time (sec):" + "%6.0f" % (time.time() - tic))
result = mean_squared_error(y_test, LGB.predict(X_test))
```

```python
LGB.fit(X_train,y_train)
test_score=LGB.score(X_test,y_test)
train_score=LGB.score(X_train,y_train)
y_pred_lgb=LGB.predict(X_test)
print("\nThe test score is:" ,test_score)
print("\nThe train score is:" ,train_score)
```
