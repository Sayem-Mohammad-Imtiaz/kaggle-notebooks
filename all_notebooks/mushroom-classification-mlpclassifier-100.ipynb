```python
# Importing the libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
```

```python
# Reading the dataset

data = pd.read_csv('../input/mushroom-classification/mushrooms.csv')
data.head()
```

```python
# Exploring the dataset

data.describe()
```

```python
# Check null values

data.isnull().sum(axis=1)
```

```python
# One hot encoding all the variables

k = data.keys()
for i in k:
    le = LabelEncoder()
    n = str(i) + "_n"
    data[n] = le.fit_transform(data[i])

for i in k:
    del data[i]
data.head()
```

```python
# Defining dependant and independent variables

X = (data.drop("class_n", axis=1)).values
y = (data["class_n"]).values
```

```python
# spliting data in train/test sets and applying MLP
# Using cross validation method for spliting, and GridsearchCV to optimize the hyper-parameters.

kfold_f = StratifiedKFold(n_splits=3, shuffle=True, random_state=1)
score = []
for i, (train_index, test_index) in enumerate(kfold_f.split(X, y)):
    x_train, x_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    kfold = StratifiedKFold(n_splits=3, shuffle=True, random_state=1)
    clf = MLPClassifier()
    param_grid = {"hidden_layer_sizes": np.array([70, 80, 90, 100, 150, 200])}

    gridsearch = GridSearchCV(
        estimator=clf, param_grid=param_grid, scoring="accuracy", n_jobs=10, refit=True
    )

    gridsearch.fit(x_train, y_train)

    score.append(gridsearch.score(x_test, y_test))

```

```python
Accuracy = np.mean(score, axis=0)
Accuracy * 100
```
