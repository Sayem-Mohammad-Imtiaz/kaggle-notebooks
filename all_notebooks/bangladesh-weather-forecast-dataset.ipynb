```python
# Importing "numpy" & "pandas" library
import numpy as num
import pandas as pd
```

```python
df=pd.read_csv('../input/bangladesh-weather-dataset/Temp_and_rain.csv')
df
```

```python
y = df['rain']
x = df[['tem','Month']]
```

```python
import sklearn.model_selection as model_selection
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(x,y,test_size=0.2, random_state = 200)
```

```python
len(x_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg_Model = LinearRegression()
```

```python
reg_Model.fit(x_train,y_train)
```

```python
train_score = reg_Model.score(x_train,y_train)
test_score = reg_Model.score(x_test,y_test)

print(f'Train Accuracy : {train_score:.3f}')
print(f'Test Accuracy : {test_score:.3f}')
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knr1 = KNeighborsRegressor(n_neighbors=5)
```

```python
knr1.fit(x_train,y_train)
```

```python
# Checking train accuracy & test accuracy
train_score = knr1.score(x_train,y_train)
test_score = knr1.score(x_test,y_test)

print(f'Train Accuracy : {train_score:.3f}')
print(f'Test Accuracy : {test_score:.3f}')
```
