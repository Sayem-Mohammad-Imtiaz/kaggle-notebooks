Let's look at the employees who have been working for more than 2 years at the company.  Most of the employees seems to stay in the same job role.

```python
a = data.YearsInCurrentRole[data.YearsAtCompany>2]   #a and b includes employees who work at the company for more than 2 years
b = data.YearsAtCompany[data.YearsAtCompany>2]
z = list(map(lambda x, y: x/y, a,b))

i=0
c1=0
c2=0
while(i<len(z)):
    if (z[i]<=0.5):
        c1 = c1 + 1
    elif (z[i]>0.5):
        c2 = c2 + 1
    i = i + 1
    
c1,c2
```
