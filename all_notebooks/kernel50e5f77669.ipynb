```python
def linBF(ds):

    V = ds['Confirmed']
#print(V)

    Vlin = []


    for x in V:
        if x != 0:
            Vlin.append(math.log(x))
        else:
            Vlin.append(0)

    plt.scatter(ds.index, Vlin)


  
    

    bf = np.polyfit(ds.index, Vlin, 1)
    bf2 = np.polyfit(ds.index, Vlin, 1)
    print(bf)

    bfx = [0,len(Vlin)]
    bfy = [bf[1], bfx[1] * bf[0] + bf[1]]

#print("slope of best fit line: "+str(bf[0]))


    plt.plot(bfx, bfy)
```
