```python
import matplotlib.pyplot as plt #visualization
import seaborn as sns #modern visualization
import matplotlib.ticker as ticker
```

```python
t20=pd.read_csv("/kaggle/input/rohit-sharma-performance-dataset/T20.csv")
test=pd.read_csv("/kaggle/input/rohit-sharma-performance-dataset/ODI.csv")
odi=pd.read_csv("/kaggle/input/rohit-sharma-performance-dataset/TEST.csv")
```

```python
X=t20['Year']
Y=t20['Runs']
ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))


plt.plot(X, Y, c = 'k')
plt.show()
```

```python
X=test['Year']
Y=test['Runs']
ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))

plt.bar(X, Y)
plt.show()
```

```python
X3=odi['Year']
Y3=odi['Runs']

plt.scatter(X3, Y3)
plt.show()
```

```python
X = t20['Year']
four =t20['4s'] 
six = t20['6s']

plt.figure(figsize=(8,8))
X_axis = np.arange(len(X))
ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(2))

plt.bar(X_axis - 0.2, four, 0.4, label = '4s')
plt.bar(X_axis + 0.2, six, 0.4, label = '6s')

plt.xticks(X_axis, X)
plt.xlabel("Year")
plt.ylabel("Total No of Boundary")
plt.title("Number of Boundary in each Year")
plt.legend()
plt.show()
```
