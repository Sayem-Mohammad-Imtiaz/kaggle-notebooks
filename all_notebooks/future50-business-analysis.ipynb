```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
import jieba
%matplotlib inline
```

First, look at the columns in the dataset and make sure the meaning of every column.
YoY means the year over year.

For example: YOY_Sales=( 2020 Sales-2019 Sales)/2019 Sales

Then Start analyzing

```python
future50 = pd.read_csv("../input/restaurant-business-rankings-2020/Future50.csv")
future50.head()
```

```python
#餐廳的Sales都在20-50之間
plt.figure(figsize=(12,8))
plt.bar(future50.Restaurant, future50.Sales)
plt.xticks(rotation='vertical')
plt.ylabel('Sales', fontsize=12)
plt.xlabel('Restaurant', fontsize=12)
plt.title('2020 Restaurant\'s Sales')
plt.show()
```

All of the Restaurant's Sales in 2020 is between 20-50

```python
plt.figure(figsize=(12,8))
plt.bar(future50.Restaurant, future50.new_YOY_Sales)
plt.xticks(rotation='vertical')
plt.ylabel('YOY_Sales(%)', fontsize=12)
plt.xlabel('Region', fontsize=12)
plt.title('Sales in each Restaurant')
plt.show()
```

Due to Covid-19, most of the restaurant's YOY Sales are below 40% 

```python
#location1.reset_index(inplace=True)
#del location1['Rank']
future50['City'] = future50['Location'] .map(lambda x:x.split(',')[0])
future50['Region']= future50['Location'] .map(lambda x:x.split(',')[1])
future50['Region']= future50['Region'].str.replace('  Calif.',' Calif.')#原本出現了重複資料
region_ = future50.groupby('Region').sum()
region_.reset_index(inplace=True)
region_
```

[Data Cleaning]:

1. Seperate the column, make the data in Location to City and Region, so we can find that which Rigion in America has most Sales.

2. If we directly groupby Region might find that there are two Calif. in the result but actually we only need one.
Then we soon find out that there are spaces in front of Calif., so we need to fix it.


```python
#plt.figure(figsize=(12,8))
plt.bar(region_.Region, region_.Sales)
plt.xticks(rotation='vertical')
plt.ylabel('Sales', fontsize=12)
plt.xlabel('Region', fontsize=12)
plt.title('Sales in each Region')
plt.show()
```

we find that Sales Ranking: Calif. > N.y. > Ohio

```python
future50.groupby(['Region','Restaurant']).sum()
```

Then try to figure out why the Sales is higher than other Region in America,

we soon find out becuase there are 10 restaurants in Calif., 8 restaurants in N.Y., 4 restaurants in Ohio




```python
future50['new_YOY_Sales'] = future50['YOY_Sales'] .map(lambda x:x.split('%')[0]).astype(float)
region__ = future50.groupby('Region').new_YOY_Sales.sum()
region__= pd.DataFrame(region__)
region__
#future50.dtypes
plt.bar(region__.index, region__.new_YOY_Sales)
plt.xticks(rotation='vertical')
plt.ylabel('YOY_Sales(%)', fontsize=12)
plt.xlabel('Region', fontsize=12)

plt.title('YOY Sales in each Region')
plt.show()
```

Although Sales Ranking: Calif. > N.y. > Ohio

YOY Sales Ranking: Calif. > N.C. > N.y.

The interesting thing is although the Reataurant Sales in Washi. is low , the YOY Sales is related High.



```python
#不論是否為加盟店，年增長銷售額與年增長量皆成正比
plt.figure(figsize=(12,8))
sns.scatterplot(future50.YOY_Units, future50.new_YOY_Sales,hue= future50.Franchising)
```
