```python
import pandas as pd 
import matplotlib.pyplot as plt 
import os 
import numpy as np
import seaborn as sns 
from sklearn import preprocessing 
```

```python
df = pd.read_csv('../input/suicide-rates-overview-1985-to-2016/master.csv')
df_copy = df.copy()
```

#### the dataframe consists of 12 columns 
      Country : Column contain name of the country 
      
      Year : The Yearin which data was conducted, the data run from 1987 to 2020
     
      age : age group

      Sex : Gender of the case Male and Female 
      
      suicides_no : number of suicides 
      
      Population :  number of citizens 
      
      suicides/100k pop : rate of suicide per 100k population 
      
      HDI : Human development Index
      
      country-year : country appended with year
     
      gdp_for_year   : the gross domestic product per year
       
      gdp_per_capita : the gross domestic product per person 
      
      generation : (Generation X, Silent, G.I Generation, Boomers, Generation Z)

```python
countries = np.unique(df['country'])
Age_group = np.unique(df['age'])
print('Countries :')
print(countries)
print('Age Groups :')
print(Age_group)
```

### Encoding the Gender column into 0 and 1 instead of Female and Male to spot the correlation among suicide and gender

```python
label_encoder = preprocessing.LabelEncoder() 
df_copy['sex'] = label_encoder.fit_transform(df_copy['sex'])
```

#### we will encoding using one hot encoding technique for age variable since it's an ordinal one we have to maintain a certain order

```python
df_copy = pd.get_dummies(data = df_copy, columns=['age'])
```

```python
df_copy
```

```python
sns.heatmap(df_copy.corr())
```

```python
sns.catplot(x="sex", y="suicides/100k pop", hue="sex",
            palette={"male": "b", "female": "r"},
            markers=["*", "o"],
            kind="point", data=df)
```

```python
sns.catplot(x="sex", y="suicides_no", hue="sex",
            palette={"male": "b", "female": "r"},
            markers=["*", "o"],
            kind="point", data=df)
```

```python
from scipy.stats import pearsonr
corr, _ = pearsonr(df_copy['sex'],df['suicides/100k pop'])
print('Pearsons correlation between sex and suicide/100k pop: %.1f' % corr)

corr, _ = pearsonr(df_copy['sex'],df['suicides_no'])
print('Pearsons correlation between sex and suicide number: %.2f' % corr)
```

##### The graph below shows the age group which has the highest suicide_rate

```python
plt.figure(figsize=(8,4))
ax = sns.catplot(x="age", y="suicides_no", hue="sex",
            palette={"male": "b", "female": "r"},
            markers=["o", "o"],
            kind="point", data=df)
plt.tight_layout()
```

##### now we will calculate the pearson correlation between age and suicide number

##### highest suicide age is between 35-54 and as previous graph showed, men suicide rate is higher than women

```python
for i in df_copy:
    if 'age' in i:
        corr, _ = pearsonr(df_copy[i],df['suicides_no'])
        print('Pearsons correlation between ' + i + ' and suicide/100k pop: %.1f' % corr)
```
