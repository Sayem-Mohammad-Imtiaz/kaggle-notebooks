# Workflow Overview
The data journalism workflow I follow is depicted as follows. 
Before we begin with any exploration or visualizations of the data, we first have to collect, clean and preprocess the data (see [Lesson 1: Data Preparation](https://www.kaggle.com/iamleonie/data-journalism-data-preparation-1-5)). 
Then we get knee deep into the data and explore it (see [Lesson 2: Exploratory Data Analysis](https://www.kaggle.com/iamleonie/data-journalism-exploratory-data-analysis-2-5)).
Once we are done with our research, we have gathered a lot of unstructured information which we will order into a story concept (see [Lesson 3: Story Concept](http://www.kaggle.com/iamleonie/data-journalism-story-concept-3-5)).
Next we will create the matching visualization to the message or story we want to convey to our audience (see [Lesson 4: Visualization & Storytelling](https://www.kaggle.com/iamleonie/data-journalism-visualization-storytelling-4-5)).
Finally, we will put it all together to a comprehensive analytics report (see the [final result](https://www.kaggle.com/iamleonie/data-journalism-visualization-storytelling-4-5/))
