```python
input = pd.read_csv('/kaggle/input/forest-fires-in-brazil/amazon.csv', encoding = 'latin1')
```

```python
input['month'].unique()
```

```python
month_map = {'Janeiro' : 'January', 'Fevereiro':'February','Mar√ßo':'March','Abril':'April','Maio':'May','Junho' : 'June', 'Julho' : 'July', 'Agosto' : 'August', 'Setembro':'September','Outubro':'October', 'Novembro':'November', 'Dezembro':'December' }
input['month'] = input['month'].map(month_map)
input.month.unique()
```

```python
plt.figure(figsize=(18,10))

ax = sns.lineplot(x = 'year', y = 'number', data = input, estimator = 'sum', color = 'orange', lw = 3, err_style = None)

plt.title('Total Fires in Brazil : 1998 - 2017', fontsize = 14)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Number of Fires', fontsize = 14)

ax.xaxis.set_major_locator(plt.MaxNLocator(19))
ax.set_xlim(1998, 2017)

ax.get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, p: format(int(x), ',')))
```

Convert Latin month names to English Name

```python
fire_state_data= input[input['state'].isin(['Amazonas','Mato Grosso','Paraiba','Sao Paulo','Rio'])].groupby(by = ['year','state', 'month']).sum().reset_index()
plt.figure(figsize=(18,10))
flp = sns.lineplot(x='year', y='number', data= fire_state_data, lw = 3, estimator = 'sum', err_style= None, hue= 'state', palette= "Set1")
plt.title('Total Fires in Brazil by State : 1998 - 2017', fontsize = 18)
plt.xlabel('Year', fontsize = 15)
plt.ylabel('Number of Fires', fontsize = 15)
flp.xaxis.set_major_locator(plt.MaxNLocator(19))
flp.set_xlim(1998, 2017)
```

```python
df_state = input.groupby("state").agg({"number":"sum"}).sort_values("number", ascending=False).reset_index()
plt.figure(figsize = (18,8))
plt.title('State Wise Fire',color = 'white',fontsize = 15)
sns.barplot(x="state", y="number", data=df_state)
plt.xticks(rotation=75,color = 'white',fontsize = 15)
plt.yticks(color = 'white',fontsize = 15)
plt.xlabel('State',color= 'yellow',fontsize = 15)
plt.ylabel('Number',color= 'yellow', fontsize = 15)
plt.show()
```

****At the top Mato Grosso state and the at bottom srgipe****

```python
fire_mato_grosso = input[input['state'].isin(['Mato Grosso'])].groupby(by = ['month']).sum().sort_values(, ascending=False).reset_index()
fire_remove_year = fire_mato_grosso.drop(columns = ['year'])

fire_remove_year

```

```python
plt.figure(figsize =(12,4))
plt.title('Month Wise Fire in Moto Grosso State',color = 'white',fontsize = 15)
sns.barplot(x="month",order = ['January','February','March','April','May','June','July','August','September','October','November','December'], y="number", data=fire_remove_year)
plt.xticks(rotation=75,color = 'white')
plt.yticks(color = 'white')
plt.xlabel('Month',color= 'yellow')
plt.ylabel('Number',color= 'yellow')
```

****
From the above graph we can see that most number of fire happened in months of January, March, April, May, October and November while least one in Septemeber.
