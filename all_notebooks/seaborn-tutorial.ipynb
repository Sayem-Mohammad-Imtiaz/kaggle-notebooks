# Seaborn Tutorial

## Seaborn vs Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('classic')
```

```python
%matplotlib inline
```

```python
rng = np.random.RandomState(0)
```

```python
x = np.linspace(0, 10, 250)
```

```python
y = np.cumsum(rng.randn(250, 6), 0)
```

```python
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='best')
```

```python
import seaborn as sns
```

```python
sns.set()
```

```python
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left')
```

## Exploring Seaborn Plots

```python
iris = pd.read_csv("../input/iris/Iris.csv")
```

```python
iris.head()
```

```python
setosa=iris.loc[iris.Species== "Iris-setosa"]
virginica = iris.loc[iris.Species == "IrÄ±s-virginica"]
```

## Histogram and Density Plots 

```python
setosa.SepalLengthCm.plot.hist()
```

```python
sns.kdeplot(
    setosa.SepalLengthCm, shade=True, color="r")
```

```python
sns.histplot(iris.SepalLengthCm, kde=True)
```

```python
sns.histplot(
    iris.SepalLengthCm, kde=True, color="r")
sns.histplot(
    iris.SepalWidthCm, kde=True,color="b")
```

## Two-Dimensional Graphics

```python
sns.kdeplot(
    data=iris, x="SepalLengthCm", y="SepalWidthCm")
```

```python
sns.kdeplot(
    data=iris,
    x="SepalLengthCm", 
    y="SepalWidthCm", 
    shade=True)
```

```python
with sns.axes_style('white'):
    sns.jointplot(
        data= iris, 
        x="SepalLengthCm", 
        y="SepalWidthCm", 
        kind='kde')
```

```python
with sns.axes_style('white'):
    sns.jointplot(
        data= iris, 
        x="SepalLengthCm", 
        y="SepalWidthCm", 
        kind='hex')
```

```python
sns.pairplot(iris, hue='Species')
```

```python
tips=pd.read_csv('../input/tipping/tips.csv')
```

```python
tips.head()
```

```python
tips["tip_percent"]=tips[
    "tip"]*100/tips["total_bill"]
```

```python
grid=sns.FacetGrid(
    tips, row="smoker",col="time", 
    margin_titles=True)
```

```python
grid.map(
    plt.hist, "tip_percent", 
    bins=np.linspace(0,40,15))
```

## Box Plots

```python
with sns.axes_style(style='ticks'): 
    g = sns.catplot(x="day", y="total_bill", 
                    data=tips, kind="box")
g.set_axis_labels("Day", "Total Account")
```

```python
sns.jointplot(x="total_bill", 
              y="tip", 
              data=tips, 
              kind='reg')
```

## Bar Plots

```python
with sns.axes_style('white'):
	g = sns.catplot(x="size", data=tips, 
                    aspect=2,kind="count", color='steelblue') 	
	g.set_xticklabels(step=5)
```
