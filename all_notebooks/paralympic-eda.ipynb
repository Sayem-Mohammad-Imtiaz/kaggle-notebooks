```python
po2020.head(9).sort_values(by="rank by total",
                        ascending=True).style.set_properties(**{'background-color': '#F0F0F0',
                                    'color': '#222222',
                                    'border': '1.5px  solid black'}).bar(color='#FEC63D', 
                                                                         vmin=100_000, 
                                                                         subset=['Gold']).bar(color='#C2C2C2', 
                                                                                                    vmin=100_000, 
                                                                                                    subset=['silver']).bar(color='#CE6A2E', 
                                                                                                                                    vmin=100_000, 
                                                                                                                                    subset=['bronze']).hide_index()
```

```python
po2020bytotal[['Team/Npc','bronze','silver','Gold','rank by total']].iloc[:10].style.set_caption('Medals by Country: Summer Olympic Games sorted by Gold Medals [Top 15]')\
.bar(subset=['Gold'], color='#FEC63D')\
.bar(subset=['silver'], color='#C2C2C2')\
.bar(subset=['bronze'], color='#CE6A2E')\
.hide_index()
```

```python
plt.figure(figsize=(18,10))
sns.color_palette('YlGn', len(po2020))
sns.barplot(x='Team/Npc',y='Gold',data=po2020[:10])
path = mpimg.imread('../input/medals/gold.png')
imagebox = OffsetImage(path , zoom=1.6)
```

```python
#plot
plt.figure(figsize=(27,15));
ax = sns.barplot(x='Team/Npc',y='Gold',data=po2020[:10])

ax.set_xlabel("Total",fontsize=20, weight='bold')
ax.set_ylabel("Country",fontsize=20, weight='bold')
ax.tick_params(labelsize=10, width=0.5, length=1.5)

plt.title("Top 30 Countries with Medals",size=20,weight='bold')  

#image
path = mpimg.imread('../input/medals/gold.png')
imagebox = OffsetImage(path , zoom=0.4)
xy = (0.5, 0.7)
ab = AnnotationBbox(imagebox, xy, frameon=False, pad=1, xybox=(100.5, 16))

ax.add_artist(ab)
```
