```python
bank=pd.read_csv("../input/credit-card-customers/BankChurners.csv")
```

```python
bank.head()
att=bank[bank['Attrition_Flag']=='Attrited Customer']

```

```python
import plotly.express as px

fig = px.histogram(att, x="Customer_Age",nbins=5)
fig.show()
```

```python
att=att[np.logical_and(att["Customer_Age"]>30 , att["Customer_Age"]<60)]
```

```python
att.shape
```

```python

att=att[att['Marital_Status']!='Unknown']
att=att[att['Education_Level']!='Unknown']
att=att[att['Income_Category']!='Unknown']

```

```python
att.head()
```

```python
att.Income_Category.unique()
```

```python
fig = px.sunburst(att, path=['Marital_Status', 'Education_Level'], values='Credit_Limit')
fig.show()
```

```python






import plotly.express as px


fig = px.density_contour(att, y="Total_Trans_Amt", x="Dependent_count")
fig.update_traces(contours_coloring="fill", contours_showlabels = True)
fig.show()
```

```python
import plotly.express as px

fig = px.scatter(att, x="Credit_Limit", y="Total_Trans_Amt", hover_name="Dependent_count", color="Income_Category",
           size="Total_Revolving_Bal")
fig.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure()

fig.add_trace(go.Scatter(x=att.Customer_Age, y=att.Credit_Limit,
                    mode='markers',
                    name='lines+markers'))


fig.show()
```

```python
import plotly.express as px



fig = px.box(att, x="Dependent_count", y="Customer_Age")
fig.update_traces(quartilemethod="exclusive") # or "inclusive", or "linear" by default
fig.show()
```
