# **Removing Outliers Using Percentile Method**

```python
import pandas as pd
import numpy as np

df = pd.read_csv('ab-nyc_2019.csv')
df.head()

df.shape #Showcase the size of the datasets (rows, columns)

df.columns 

upper_bound = df['price'].quantile(0.9995)
print ("Upper_bound Value:", upper_bound)
lower_bound = df['price'].quantile(0.0005)
print ("Lower-bound Value:", lower_bound)
max_price = max(df['price'])
print ("Max price:", max_price)
min_price = min(df['price'])
print ("Min price:", min_price)

print (f'Number of points having price greater than {upper_bound}:', len(df[df['price']>upper_bound]))

print (f"Number of points having price lower than {lower_bound}:", len(df[df['price']<lower_bound]))

df_percentile = df[(df['price']<upper_bound) & (df['price']>lower_bound)]
df_percentile 

df_percentile.shape

df.shape
```
