># BEST SCORE FROM HOUSING PRICES COMPETITION
># Defining a function to get the MAE and using a FOR repetition to check all possibles leaf_nodes and respective MAE values, then getting the best score from the max_leaf_nodes setted. After that checking all the data X,y with the best score found.

># Importing panda and numpy as common to get the dataset and modify the type of the number.

```python
import pandas as pd 
import numpy as np 
house_file = '../input/home-data-for-ml-course/train.csv'
house_data = pd.read_csv(house_file)
```

># Using a FOR repetition to deleting the columns with NaN 

```python
for columns in house_data:
    if house_data[columns].isnull().sum() >1000:
        house_data = house_data.drop(columns, axis = 1)
```

># Replacing some NaN that still exist in the dataframe for 0.

```python
house_data.fillna(0, inplace= True)
```

># Choosing the parameters X and y for the dataset and converting the type float64 to float32

```python
y = house_data.SalePrice.astype(np.float32, copy= False)
features = ['LotFrontage', 'LotArea', 'MoSold','YrSold']
X = house_data[features].astype(np.float32, copy = False)
```

># Importing the libs for training the dataset and get the MAE

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error 
```

># Training the dataset with train_test_split that get random trains and test subsets

```python
train_X, val_X, train_y, val_y = train_test_split(X,y, random_state=1)
```

># Defining a function that will return the MAE value for each X, y splitted and the max_leaf_nodes setted.

```python
def get_mae(max_leaf_nodes, train_X, val_x, train_y, val_Y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    predict_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, predict_val)
    return (mae)
```

># By some leaf_nodes defined, the FOR repetition will pass through all the values and set in the function get_mae, to get each MAE value for each possible leaf

```python
possible_max_leaf_nodes = [5, 25, 50, 100, 250, 500]
for max_leaf_nodes in possible_max_leaf_nodes:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print(max_leaf_nodes, my_mae)
```

># FOR repetition to get the best score from the possible_max_leaf_nodes

```python
score = {leaf_size: get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y) for leaf_size in possible_max_leaf_nodes}
best_size_nodes = min(score, key= score.get)
print(best_size_nodes)
```

># Testing the model for the whole X,y dataset with the best score got in the code above, now changing the DecisionTreeRegressor to a specific value 'best_size_nodes'

```python
house_final_model = DecisionTreeRegressor(max_leaf_nodes= best_size_nodes, random_state = 0)
house_final_model.fit(X, y)
```

># Predicting the MAE now for whole dataset not only for val_X, val_y

```python
predictions_final = house_final_model.predict(X)
val_mae = mean_absolute_error(y, predictions_final)
print(val_mae)
```
