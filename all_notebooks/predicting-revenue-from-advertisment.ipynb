```python
df = pd.read_csv(
    os.path.join(root, 'Test_Data.csv'))
#df.isnull().any()[df.isnull().any()>0]
#df = df.fillna(df.mean)
df
```

```python
#Checking null values
df.isnull().sum()
#No null in the dataset
```

```python
#Checking if datasets have duplicates values
df2 = df.drop_duplicates()
print("Dataset before duplicates {} and after duplicate deletion {}".format(df.shape, df2.shape))
#No duplicates
```

```python
#Checking all unique elements
df2['campaign'].unique()
df2['adgroup'].unique()
#Campaign has 1 dataset and adgroup has 4 dataset, Campaign could be dropped
df2.drop(['campaign'], axis = 'columns')
```

```python
#Convert adgroup and ad string into integer
df2['adgroup'] = df2['adgroup'].apply(lambda x : int(x.split(' ')[1]))
df2['ad'] = df2['ad'].apply(lambda x : int(x.split(' ')[1]))
df2['ad']
```

```python
#Preliminary analysis, understanding data
df_group = df2.groupby(['adgroup', 'ad']).sum().sort_values(by ='conversions', ascending = False).sort_index()
df_group = df_group['conversions']
df_group
```
