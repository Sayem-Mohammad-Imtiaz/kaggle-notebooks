```python
df = pd.read_csv('/kaggle/input/pakistan-car-prices/updated_pakwheels.csv')
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.drop(['Ad No','URL'],axis=True, inplace=True)
```

```python
df.head()
```

```python
df['Brand Name'] = df.Name.apply(lambda x : x.split()[0])
```

```python
df['Location 2'] = df.Location.apply(lambda x: x.split()[-1])
```

```python
def get(val):
    if val!= 'Un-Registered':
        return 'Registered'
    else:
        return val
df['Registered'] = df['Registered City'].apply(get)
```

```python
df.head(3)
```

```python
def Engine(val):
    dis = {
        1:"Petrol",
        2:'Diesel',
        3:'Hybrid'   
    }
    return dis[val]
def Trans(val):
    dis = {
        1: "Automatic",
        2: "Manual"
    }
    return dis[val]
```

```python
df['Transmision Type'] = df.Transmission.apply(Trans)
df['Engine'] = df['Engine Type'].apply(Engine)
```

```python
df.head()
```

```python
df2  = df.drop(['Name','Location','Registered City','Engine Type','Transmission','Last Updated'],axis=1)
```

```python
df2.head()
```

```python
# 1 for Hatchback 2 for Sedan 3 for Suv 4 for mini van 5 for Crossover 6 for Van
```

```python
def Body(val):
    dist = {
        1:"Hatchback",
        2:"Sedan",
        3:'SUV',
        4:'Mini Van',
        5:'Cross Over',
        6:'Van'
    }
    return dist[val]
```

```python
df2['Body'] = df2['Body Type'].apply(Body)
```

```python
df2.head()
```

```python
df2.drop('Body Type',axis=1,inplace=True)
```

```python
df2.head()
```

```python
df2.isnull().sum()
```

```python
corr = df2.corr()
```

```python
corr.loc['Price']
```

```python
df2.drop('Color',axis=1,inplace=True)
```

```python
df2.drop(['Features'],axis=1,inplace=True)
```

```python
df2.head()
```

```python
new = pd.DataFrame(df2.dtypes,columns=['dtype'])
```

```python
list(new[new['dtype']=='object'].index)
```

```python
def OneHotEncoder(data):
    values = pd.get_dummies(data)
    return values.drop(list(data.value_counts().tail(1).index),axis=1)
```

```python
df3 = df2.copy()
```

```python
for i in list(new[new['dtype']=='object'].index):
    df3= pd.concat([df3,OneHotEncoder(df3[i])],axis=1)
```

```python
df3.head()
```

```python
df4 = df3.drop(list(new[new['dtype']=='object'].index),axis=1)
```

```python
df4.head()
```

```python
df4.isnull().sum()
```

```python
X = df4.drop('Price',axis=1)
Y = df4.Price
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y,random_state=1)
```

```python
model1 = RandomForestRegressor()
model2 = LinearRegression()
```

```python
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
model2.score(x_test,y_test)
```

```python
# Linear Regression Model is not Good enough to go
```

```python
RFM = RandomForestRegressor(n_estimators=50)
RFM.fit(x_train,y_train)
```

```python
RFM.score(x_test,y_test)
```
