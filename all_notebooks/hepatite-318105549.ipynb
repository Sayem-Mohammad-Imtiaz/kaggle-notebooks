### Crie um histograma da distribuição de idades (age)

```python
import matplotlib.pyplot as plt
plt.hist(df['age'], bins=None, rwidth=0.9 )
plt.title('Distribuição de Idades')
plt.xlabel('idade')
plt.ylabel('contagem')
plt.show
```

### Separe o dataset nos atributos de entrada X e objetivo y

```python
X = df.loc[:,'age':'histology']
y = df['class']
```

### Utilize a biblioteca de pré-processamento para normalizar os atributos de entrada (X)

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X = ss.fit_transform(X)
```

### Separe o dataset em Treino e Teste

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

### Crie o modelo para MLP, treine-o e faça a predição com os dados de teste (y_pred)

```python
from sklearn.linear_model import Perceptron
pp = Perceptron()
modelo = pp.fit(X_train, y_train)
y_pred = modelo.predict(X_test)
```

### Calcule o Score do Modelo

```python
score = modelo.score(X,y)
score
```
