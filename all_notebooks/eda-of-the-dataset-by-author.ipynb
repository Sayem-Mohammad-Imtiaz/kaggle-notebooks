```python
data = pd.read_csv('../input/goodreads-quotes/quotes.csv')
```

```python
data = data.drop_duplicates().reset_index()
```

```python
data.to_csv('quotes.csv', index=False)
```

Analyze the distribution of words in the inspirational quotes.

```python
concat_quotes = ' '.join([i for i in data.quote.astype(str)])
print(concat_quotes[:1000])
```

```python
import stylecloud

stylecloud.gen_stylecloud(text=concat_quotes,
                          icon_name='fas fa-lightbulb',
                          palette='colorbrewer.diverging.Spectral_11',
                          background_color='black',
                          gradient='horizontal',
                          size=1024)
```

```python
import os
from IPython.display import Image
Image(filename="./stylecloud.png", width=1024, height=1024)
```
