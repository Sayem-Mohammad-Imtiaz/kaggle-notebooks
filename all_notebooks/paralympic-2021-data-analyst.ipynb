## Data Analysis

Now that our data is clean, we can start do analysis with the data provided in the data frame.

It is time to import some packages provided by python. 

```python
## import packages that can be use for data analysis and visualization
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

Here I want to analyse the ranking of team in paralympics 2021.

```python
rank=df1.sort_values(by='Rank by total medals',ascending=True)
rank.head(20)
```

let's see how many countries/team from each continent

we can also see the total of gold, silver & bronze medals obtained during the paralympic 2021

```python
count = df1[['Gold','Silver','Bronze']].sum().reset_index().rename(columns={'index':'Medal',0:'Total obtained'})
print(count)
```

It is time to plot some data visualization to do further analysis. 

```python
vari=['Team/Npc','Rank by total medals','Total Medals']
byRank=df1.sort_values('Rank by total medals', ascending=True).reset_index().head(20)
byRank=byRank[vari].head(20)

plt.figure(figsize = (15,8))
bar2=sns.barplot(y='Team/Npc',x='Total Medals', data=byRank)
bar2.set_title('Team/Npc against Total Medals')
```

```python
df1['Continent'].value_counts()
```

```python
plt.figure(figsize = (10,8))
bar1=sns.countplot(x='Continent', data=df1,palette=['#6F8FAF'])
bar1.set_title('Figure 2: Number of Team/NPC in Continent',fontsize=15)
bar1.set_xlabel('Continent',fontsize=12)
bar1.set_ylabel('Number of Team/Npc')
bar1.tick_params(axis='x', rotation=45)
```

```python
dfcont=df1.groupby('Continent', as_index=False).agg({"Total Medals": "sum","Gold":"sum","Silver":"sum","Bronze":"sum"})
dfcont
```

```python
plt.figure(figsize = (10,8))
bar3=sns.barplot(x='Continent',y='Total Medals',data=dfcont,palette=['#6F8FAF'])
bar3.set_title('Figure 3: Total medals obtained by each Continents',fontsize=15)
bar3.tick_params(axis='x', rotation=45)
```

Based on both Figure 2 and Figure 3, we can see Europe continent has the highest value in both; team and total medal obtained. As Europe Team has the largest number of team, that would tell us that Europe will obtained highest TOTAL number of medals. eventhough People's Republic of China has ranked as no 1 in the list. However comparing with the continent, Europe still obtained highest number of medals. 

```python
plt.figure(figsize=(16,7))
xpos = np.arange(len(dfcont['Continent']))
plt.xticks(xpos,dfcont['Continent'],rotation=90,fontweight='bold',fontsize=13)
plt.bar(xpos-0.3,'Gold',data=dfcont,color='gold',width=0.3)
plt.bar(xpos,'Silver',data=dfcont,color='#C0C0C0',width=0.3,label='Silver')
plt.bar(xpos+0.3,'Bronze',data=dfcont,color='#C39B77',width=0.3,label='Bronze')
plt.yticks(fontsize=12,color='black',fontweight='bold')
plt.legend(labels=["Gold","Silver","Bronze"])
plt.show()
```
