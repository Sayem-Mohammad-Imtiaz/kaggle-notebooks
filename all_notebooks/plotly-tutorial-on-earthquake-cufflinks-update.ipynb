We have values the position and magnitude of earthquakes. And measurements of solar system planets.

## Scatter Plot of Magnitude and Rectascention of Sun

```python
trace = go.Scatter(
    x = df["Sun.rectascension"],
    y = df["earthquake.mag"],
    mode='markers'
)

layout = go.Layout(
    title='Correlation between magnitude and Rectascension',
    xaxis=dict(title='Rectascension'),
    yaxis=dict(title='Magnitude'),
    showlegend=True    
)

fig = dict(data=[trace], layout=layout)
offline.iplot(fig)
```

```python
## idem with cufflinks
df.iplot( x='Sun.rectascension', y='earthquake.mag', 
          kind='scatter', mode='markers', colors='blue', size=5,
          xTitle='Rectascension',
          yTitle='Magnitude',
          title="Correlation between magnitude and Rectascension",)
```

## Transition of Magnitude in these 20years

```python
df = df.sort_values(by='earthquake.time')

trace = go.Scatter(
    x = df["earthquake.time"],
    y = df["earthquake.mag"],
    mode='markers+lines'
)

layout = go.Layout(
    title='Magnitude in these 20years',
    xaxis=dict(title='time'),
    yaxis=dict(title='Magnitude'),
    showlegend=True    
)

fig = dict(data=[trace], layout=layout)
offline.iplot(fig)
```

```python
## idem with cufflinks
df.sort_values(by='earthquake.time')\
  .iplot( x='earthquake.time', y='earthquake.mag', 
          kind='scatter', mode='lines+markers', colors='blue', size=5,
          xTitle='Time',
          yTitle='Magnitude',
          title='Magnitude in these 20years',
        )
```

## Histogram of Magnitude

```python
trace = go.Histogram(x=df["earthquake.mag"])

layout = go.Layout(
    title='Magnitude Distribution',
    xaxis=dict(title='Magnitude'),
    yaxis=dict(title='Frequency')
)

fig = dict(data=[trace], layout=layout)
offline.iplot(fig)
```

```python
## idem with cufflinks
df[ 'earthquake.mag'].\
   iplot( kind='histogram', bins=50, colors='blue',
          xTitle='Magnitude',
          yTitle='Frequency',
          title='Magnitude Distribution',
        )
```

## Distribution of Magnitude by each Moon Phase

```python
dsc = df[df["MoonPhase.dynamic"] == 'dsc']
asc = df[df["MoonPhase.dynamic"] == 'asc']

t1 = go.Box(
    y = dsc["earthquake.mag"].values,
    name = 'Moon Ascending phase',
    marker = dict(
        color = 'rgb(214, 12, 140)'
    )
)

t2 = go.Box(
    y = asc["earthquake.mag"].values,
    name = 'Moon Descending phase',
    marker = dict(
        color = 'rgb(0, 128, 128)'
    )
)

offline.iplot([t1, t2])
```

```python
## idem with cufflinks
dsc = df[ df[ "MoonPhase.dynamic"] == 'dsc'][ "earthquake.mag"]
asc = df[ df[ "MoonPhase.dynamic"] == 'asc'][ "earthquake.mag"]

pd.DataFrame({ 'Moon Ascending phase': asc, 'Moon Descending phase': dsc})\
   .iplot( kind='box', colors=[ 'blue', 'green'])
```

## Distribution of Earthquake

```python
t = go.Histogram2d(
    x = df["earthquake.latitude"].values,
    y = df["earthquake.longitude"].values,
    name = 'Position of Earthquake',
    opacity = 0.75    
)

l = go.Layout(
    title = 'Position of Earthquake',
    xaxis = dict(title='Latitude'),
    yaxis = dict(title='Longitude')
)

fig = dict(data=[t], layout=l)
offline.iplot(fig)
```

```python
import plotly.figure_factory as ff
y = df["earthquake.latitude"].values
x = df["earthquake.longitude"].values

colorscale = ['#7A4579', '#D56073', 'rgb(236,158,105)', (1, 1, 0.2), (0.98,0.98,0.98)]

fig = ff.create_2d_density(
    x, y, colorscale=colorscale,
    hist_color='rgb(0, 128, 222)', point_size=3,
)

offline.iplot(fig)
```
