```python
df=pd.merge(movies,ratings,on='movieId')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings=pd.DataFrame(df.groupby('title')['rating'].mean().sort_values(ascending=False))
ratings
```

```python
ratings['num of rating']=pd.DataFrame(df.groupby('title')['rating'].count())
ratings
```

```python
plt.figure(figsize=(10,4))
ratings['num of rating'].hist(bins=70)
```

```python
plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating',y='num of rating',data=ratings,alpha=0.5)
```

```python
moviemat=df.pivot_table(index='userId',columns='title',values='rating')
```

```python
JurassicPark_user_rating=moviemat['Jurassic Park (1993)']
simlar_to_JurassicPark=moviemat.corrwith(JurassicPark_user_rating)
corr_JurassicPark=pd.DataFrame(simlar_to_JurassicPark,columns=["Corelation"])
corr_JurassicPark.dropna(inplace=True)
corr_JurassicPark.sort_values("Corelation",ascending=False).head()
```

```python
corr_JurassicPark=corr_JurassicPark.join(ratings["num of rating"])
corr_JurassicPark[corr_JurassicPark["num of rating"]>100].sort_values("Corelation",ascending=False).head()
```
