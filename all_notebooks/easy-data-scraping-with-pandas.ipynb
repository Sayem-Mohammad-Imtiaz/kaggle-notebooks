## Loading the Data 

We will scrape data from a wikipedia page.
You can visit the webpage [here](https://en.wikipedia.org/wiki/Demographics_of_India).

```python
url = 'https://en.wikipedia.org/wiki/Demographics_of_India'
```

```python
dataframes = pd.read_html(url)    # load the webpage url 
```


For example, We want to extract this table.

```python
#dataframes   #all tables from the url is printed
dataframes[0:3] 
```
