```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas_profiling import ProfileReport
import seaborn as sns
```

```python
df=pd.read_csv('../input/forbes-billionaires-of-2021-20/forbes_billionaires.csv')
```

```python
profile = ProfileReport(df, title='Pandas Profiling Report')
```

```python
profile
```

```python
plt.figure(figsize=[16,4])
base_color=sns.color_palette()[0]
order=df['Country'].value_counts()[:20].index
sns.countplot(data=df,x='Country',order=order,color=base_color)
plt.title('No of Billionares over the world',fontname='Times new roman',fontsize=16)
plt.xticks(rotation=90);
```

- The United States has the highest number of billionares in the world followed by China

```python
total_money=df.groupby('Country')['NetWorth'].sum()
x=total_money.reset_index()
```

```python
x.head()
```

```python
display(x.nlargest(1,'NetWorth'))
```

```python
x2=x.sort_values('NetWorth',ascending=False)[:20]
```

```python
import plotly.express as px
fig = px.bar(x2, x='Country', y='NetWorth',title="Total amount of Billionares Money for different Countries",labels={'NetWorth':'Money in Billion $'})
fig.show()
```

```python
plt.figure(figsize=[8,4])
plt.hist(df['Age']);
plt.title('Hitogram of Billionares Age',fontname='Times new roman',fontsize=15);
plt.xlabel('Age');
plt.ylabel('Frequency');
```

```python
plt.figure(figsize=[15,4])
                    
source_order=df['Source'].value_counts().sort_values(ascending=False)[:10].index
sns.countplot(data=df,x='Source',order=source_order,color=base_color)
plt.title('Billionares Most common Money Sources',fontname='Times new roman',fontsize=18);
```

```python
plt.figure(figsize=[15,4])
status_order=df['Status'].value_counts().index
sns.countplot(data=df,x='Status',order=status_order,color=base_color);
plt.title('Billionares Marital Status');
```
