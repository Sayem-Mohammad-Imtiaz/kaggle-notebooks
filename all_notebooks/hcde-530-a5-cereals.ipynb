**Analyzing Cereal Ingredients**

Through this simple dataset, i want to explore my understanding of kaggle data visualizations as well as understand relationships between the shelf its sitting on and its rating? Which cereal manufacturer has the highest ratings?

```python
cereals = pd.read_csv("../input/80-cereals/cereal.csv")
cereals
```

**Visualization 1**


Understanding the relation between calories and ratings as many buyers prefer food products with lesser calorific content.

```python
cereals[['calories', 'rating']].plot(kind='scatter', x='calories', y='rating', alpha=0.5, c='Purple')
```

**Visualization 2**

Creating a histogram of all the numererical values of the dataset.

```python
hist=cereals.hist(figsize=(15,15))
```

**Visualization 3**

Comparison between cereals and shelf arrangements

```python
cereals['shelf'].value_counts().plot(kind='bar', y='Shelf')
```
