# Feature Engineering-How to Perform One Hot Encoding for Multi Categorical Variables

```python
import pandas as pd
import numpy as np
```

```python
# Loading the Data and using First 6 columns only.
df=pd.read_csv("../input/mercedesbenz-greener-manufacturing/train.csv",usecols=['X1','X2','X3','X4','X5','X6'])
```

```python
df.head()
```

```python
# Unique categories for each column 
for col in df.columns:
    print(col,':',len(df[col].unique()),'labels')
```

```python
# let's Examine How many columns we will obatain after one hot encoding these variables
pd.get_dummies(df,drop_first=True).shape
```

We can see that from just 6 intial categorical variable column, we end up with 117 new Variables.

What can we do instaed ?

[Winning the KDD Cup Orange Challenge with Ensemble
Selection](http://proceedings.mlr.press/v7/niculescu09/niculescu09.pdf)

In the Winning the KDD Cup Orange Challenge with Ensemble
Selection the author limit one hot encosing to the top 10 most frquent labels of the variable.This Means that they would make one Binary Variable
fpr each of the top 10 most  frequent Labels only.THis is Equivalent to
grouping  all the other labels under new category.In this case will we
dropped.Thus,the 10 new dummy variables indicate if one of the 10
most frequent labels is present [1] and not [0] for a particular
observation.

Let's do it in Python


```python
# Lets find the top 10 most frequent categories for the variable X2
df.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
# Let's Make a list with most Frequent categories of the variable.
top_10=[x for x in df.X2.value_counts().sort_values(ascending=False).head(10).index]
```

```python
top_10
```

```python
# And Now we Make The 10 Binary Variables
for label in top_10:
    df[label]=np.where(df['X2']==label,1,0)
df[['X2']+top_10].head(40)   
```

```python
# Getting whole set of dummy variables, for all the categorical Varaiable
def one_hot_encoding_top_x(data,variable,top_x_lables):
    # function to craete dummy variable for most frequent labels and we can also vary the number of
              # most frequent lables thatt we encode.
        
        for label in top_x_lables:
            df[variable+'_'+label]=np.where(df[variable]==label,1,0)

df=pd.read_csv("../input/mercedesbenz-greener-manufacturing/train.csv",usecols=['X1','X2','X3','X4','X5','X6'])    

# Encoding X2 in top 10 most frequent categories
one_hot_encoding_top_x(df,'X2',top_10)
df.head()
```

```python
# find the 10 top most frequent categories for X1
top_10=[x for x in df.X1.value_counts().sort_values(ascending=False).head(10).index]
```

```python
top_10
```

```python
one_hot_encoding_top_x(df,'X1',top_10)
df.head()
```

### Advantages




<li> StraightForward to implement.
<li>Does not Require Hours of Variable Exploration    
<li> Does Not exapnd massively the feature space (number of columns)

### Disadvantage

<li> Does Not Add any information that make the variable More Predictive.
<li> Does not kepp the information of ignored Variables.    

Because it is not unsual that categorical variables have a few dominating
categories and the remaining lables add mostly noise,this is quite
simple and straightforward approach that may be useful on many occassion.

It is nothing that every time it is to choose top 10 variables only we can choose top_5 or top_20 that mostly depends.
