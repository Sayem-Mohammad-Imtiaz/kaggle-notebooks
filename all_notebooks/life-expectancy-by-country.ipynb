```python
import os
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
data0 = pd.read_csv("../input/life-expectancy/lifeExpectancy_world.csv")
data0
```

```python
data19=data0[['Entity','2019']]
data19.sort_values(by='2019')
```

```python
data1=data0.drop('Code',axis=1)
data1
```

```python
print(list(data1['Entity']))
```

```python
data1.index=data1['Entity']
data2=data1.drop('Entity',axis=1)
data3=data2.T
data3['Year']=data3.index
data4=data3.reset_index(drop=True)
data4
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=data4['Year'],y=data4['Japan'],name="Japan"),secondary_y=False,)
fig.add_trace(go.Scatter(x=data4['Year'],y=data4['World'],name="World"),secondary_y=False,)
fig.add_trace(go.Scatter(x=data4['Year'],y=data4['United States'],name="United States"),secondary_y=False,)
fig.add_trace(go.Scatter(x=data4['Year'],y=data4['Brazil'],name="Brazil"),secondary_y=False,)
fig.add_trace(go.Scatter(x=data4['Year'],y=data4['Monaco'],name="Monaco"),secondary_y=False,)
fig.add_trace(go.Scatter(x=data4['Year'],y=data4['Central African Republic'],name="Central African Republic"),secondary_y=False,)

fig.update_layout(autosize=False,width=700,height=500,title_text="Life Expectancy by Country")
fig.update_xaxes(title_text="Date")
fig.update_yaxes(title_text="Age",secondary_y=False)
fig.show()
```
