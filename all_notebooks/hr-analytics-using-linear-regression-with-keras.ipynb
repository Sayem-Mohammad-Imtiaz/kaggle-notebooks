```python
df_dummies = pd.get_dummies(df[['Department','salary']])
```

```python
df_dummies.head()
```

```python
df.columns
```

```python
X = pd.concat([df[['satisfaction_level', 'last_evaluation', 'number_project',
        'time_spend_company', 'Work_accident', 
       'promotion_last_5years',
       'average_montly_hours_100']],df_dummies],axis=1).values
y=df['left'].values
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam,SGD
```

```python
model= Sequential()
model.add(Dense(1,input_dim=20,activation='sigmoid'))
model.compile(Adam(lr=0.5),'binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train)
```

```python
model.summary()
```

```python
y_test_pred = model.predict_classes(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
def pretty_confusion_matrix(y_true, y_pred, labels=["False", "True"]):
    cm = confusion_matrix(y_true, y_pred)
    pred_labels = ['Predicted '+ l for l in labels]
    df = pd.DataFrame(cm, index=labels, columns=pred_labels)
    return df
```

```python
pretty_confusion_matrix(y_test,y_test_pred,labels=['Stay','Left'])
```

```python
print(classification_report(y_test,y_test_pred))
```

```python
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
def build_logistic_regression_model():
    model = Sequential()
    model.add(Dense(1, input_shape=(20,), activation='sigmoid'))
    model.compile(SGD(lr=0.5),
                  'binary_crossentropy',
                  metrics=['accuracy'])
    return model

model=KerasClassifier(build_logistic_regression_model,epochs=10,verbose=0)
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
cv = KFold(5,shuffle=True)
scores = cross_val_score(model,X,y,cv=cv)
print("the cross validation accuracy is {:0.4f} Â± {:0.4f}".format(scores.mean(),scores.std()))
```
