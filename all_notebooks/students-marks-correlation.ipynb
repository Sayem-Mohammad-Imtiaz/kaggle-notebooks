```python
import numpy as np
import pandas as pd
students = pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
students.head()
```

```python
students['gender'] = students['gender'].replace(['male', 'female'], (1,0))
students['race/ethnicity'] = students['race/ethnicity'].replace(['group A', 'group B', 'group C', 'group D', 'group E'], (0, 1, 2, 3, 4))
students['parental level of education'] = students['parental level of education'].replace(["master's degree", "bachelor's degree", 'some college', 
       "associate's degree", 'high school', 'some high school'], (0,1,2,3,4,5))
students['test preparation course'] = students['test preparation course'].replace(['none', 'completed'], (0, 1))
students['lunch'] = students['lunch'].replace(['standard', 'free/reduced'], (0,1))
students.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(12,10)) 
sns.heatmap(students.corr(), annot=True, ax=ax)
```
