# Config

```python
INPUT_DIR = '../input/numerai-train-validation-with-kazutsugi-nomi/numerai_training_validation_target_nomi.csv'
```

# Read data

```python
%%time

def read_data(input_dir='../input'):
    # read data
    df = pd.read_csv(input_dir)
    
    # map features to int to reduce memory requirements
    feature_cols = df.columns[df.columns.str.startswith('feature')]
    mapping = {0.0 : 0, 0.25 : 1, 0.5 : 2, 0.75 : 3, 1.0 : 4}
    for c in feature_cols:
        df[c] = df[c].map(mapping).astype(np.uint8)
    return df

df = read_data(INPUT_DIR)
```
