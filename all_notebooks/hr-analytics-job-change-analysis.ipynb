# Context and Content

A company which is active in Big Data and Data Science wants to hire data scientists among people who successfully pass some courses which conduct by the company. Many people signup for their training. Company wants to know which of these candidates are really wants to work for the company after training or looking for a new employment because it helps to reduce the cost and time as well as the quality of training or planning the courses and categorization of candidates. Information related to demographics, education, experience are in hands from candidates signup and enrollment.

This dataset designed to understand the factors that lead a person to leave current job for HR researches too. By model(s) that uses the current credentials,demographics,experience data you will predict the probability of a candidate to look for a new job or will work for the company, as well as interpreting affected factors on employee decision.

The whole data divided to train and test . Target isn't included in test but the test target values data file is in hands for related tasks. A sample submission correspond to enrollee_id of test set provided too with columns : enrollee _id , target

## Note:

The dataset is imbalanced.
Most features are categorical (Nominal, Ordinal, Binary), some with high cardinality.
Missing imputation can be a part of your pipeline as well.

## Features
enrollee_id : Unique ID for candidate

city: City code

city_ development _index : Developement index of the city (scaled)

gender: Gender of candidate

relevent_experience: Relevant experience of candidate

enrolled_university: Type of University course enrolled if any

education_level: Education level of candidate

major_discipline :Education major discipline of candidate

experience: Candidate total experience in years

company_size: No of employees in current employer's company

company_type : Type of current employer

lastnewjob: Difference in years between previous job and current job

training_hours: training hours completed

target: 0 – Not looking for job change, 1 – Looking for a job change

# **Feature Engineering**

### Making necessary changes in various features

```python
train['city'] = train['city'].map(lambda x: x.lstrip('city_'))

train = train.dropna(subset=['experience']).reset_index(drop=True)
train['experience'] = train['experience'].map(lambda x: x.lstrip('<>'))

train['last_new_job'] = train['last_new_job'].astype(str)
for i in range(train.shape[0]):
    train.loc[i,'last_new_job'] = train.loc[i,'last_new_job'].replace('>','').replace('<','').replace('never','0').replace('nan','0')
train['last_new_job'] = train['last_new_job'].astype(int)
```

### Dropping useless columns

```python
train = train.drop(['gender', 'training_hours', 'enrollee_id'], axis=1)
```

```python
train.head()
```

### Filling the missing values with 'NA' and then changing it back to a much more appropriate value

```python
train.fillna('NA', inplace = True)
train["enrolled_university"] = train["enrolled_university"].replace('NA', 'Unknown_uni')
train["education_level"] = train["education_level"].replace('NA', 'Unknown_level')
train["company_size"] = train["company_size"].replace('NA', 'Unknown_size')
train["company_type"] = train["company_type"].replace('NA', 'Unknown_type')
train["experience"] = train["experience"].replace('NA', 0)
```

### Dummy variables

Creating Dummy varibales of various columns

```python
rel_exp = pd.get_dummies(train['relevent_experience'], drop_first = True)
enrolled_uni = pd.get_dummies(train['enrolled_university'], drop_first = True)
ed_level = pd.get_dummies(train['education_level'], drop_first = True)
maj_dis = pd.get_dummies(train['major_discipline'], drop_first = True)
comp_size = pd.get_dummies(train['company_size'], drop_first = True)
comp_type = pd.get_dummies(train['company_type'], drop_first = True)
```

### Dropping the irrelevant columns

```python
train.drop(['relevent_experience', 'enrolled_university', 'education_level', 'major_discipline', 'company_size',
                   'company_type'], axis = 1,inplace = True)
```

### Concatenating the dummy variables with the training data

```python
train = pd.concat([rel_exp, enrolled_uni, ed_level, maj_dis, comp_size, comp_type, train], axis = 1)
```

# **Model**

### Splitting the data into Training and Test set

```python
X = train.iloc[:, :-1].values
y = train.iloc[:, -1].values

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression(max_iter = 300)
log.fit(X_train, y_train)
```

```python
y_pred_log = log.predict(X_test)
print("The Accuracy Score of Logistic Regression is: {}%".format(accuracy_score(y_test, y_pred_log)*100))
print("The Confusion Matrix for Logistic Regression is: \n{}\n".format(confusion_matrix(y_test, y_pred_log)))
print('\n')
print(classification_report(y_test, y_pred_log))
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
dt.fit(X_train, y_train)
```

```python
y_pred_dt = dt.predict(X_test)
print("The Accuracy Score of Decision Tree is: {}%".format(accuracy_score(y_test, y_pred_dt)*100))
print("The Confusion Matrix for Decision Tree is: \n{}\n".format(confusion_matrix(y_test, y_pred_dt)))
print('\n')
print(classification_report(y_test, y_pred_dt))
```

### Random Forest Classifier 

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 300, criterion = 'entropy')
rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_test)
print("The Accuracy Score of Random Forest is: {}%".format(accuracy_score(y_test, y_pred_rf)*100))
print("The Confusion Matrix for Random Forest is: \n{}\n".format(confusion_matrix(y_test, y_pred_rf)))
print('\n')
print(classification_report(y_test, y_pred_rf))
```
