```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
%matplotlib inline
```

```python
df = pd.read_csv('../input/covid19-temperatureoxygenpulse-rate/qt_dataset.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.set_index('ID',inplace=True)
```

```python
df.tail()
```

```python
df.isna().sum()
```

```python
plt.style.use('fivethirtyeight')
df['Result'].value_counts().plot(kind='bar')
plt.show()
```

```python
plt.style.use('fivethirtyeight')
sns.pairplot(data=df,hue='Result')
plt.show()
```

```python
df['Result'].replace(['Positive','Negative'],[1,0],inplace=True)
```

```python
df['Result'].unique()
```

```python
x = df.drop('Result',axis=1)
y = df['Result']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#We choose kernel='rbf'
model = SVC()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred_train = model.predict(x_train)
```

```python
test = confusion_matrix(y_pred,y_test)
```

```python
plt.title('Test Data')
sns.heatmap(data=test,annot=True,fmt='g')
plt.show()
```

```python
train = confusion_matrix(y_pred_train,y_train)
```

```python
plt.title('Train Data')
sns.heatmap(data=train,annot=True,fmt='g')
plt.show()
```
