```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
import os
data = pd.read_csv(r'../input/bostonhoustingmlnd/housing.csv')
print(data)
```

```python
import torch
```

```python
inputs = data.drop('MEDV', axis = 1)    #Assigning inputs to the predictors
print(inputs)
```

```python
targets = data['MEDV']    #Assigning targets to predictor
print(type(targets))
```

```python
#We should now convert the pandas type variables into torch tensors to use PyTorch
inputs = torch.tensor(inputs.values.astype(np.float32))
targets = torch.tensor(targets.values.astype(np.float32))
```

```python
print(type(inputs), type(targets))
```

```python
print(inputs)
print(targets)
```

```python
import torch.nn as nn
from torch.utils.data import DataLoader, TensorDataset
```

```python
#We are now converting the dataset into a tensor dataset
train_ds = TensorDataset(inputs, targets)
print(train_ds[0:3])
```

```python
#If the dataset is huge, we split the datset to make it lighter for execeution
#batch_size determines the number of splits
batch_size = 10
train_dl = DataLoader(train_ds, batch_size, shuffle = True)
```

```python
model = nn.Linear(3, 1)    #Creating a Linear Regression model with 3 inputs nd 1 output  
```

```python
print(model.weight, model.bias)  
'''
When we create a Linear model, the weights and bias that need to assigned to the neural network is done automatically. 
These values are checked in the above line.
'''
```

```python
preds = model(inputs)    #Running the model to predict values
```

```python
import torch.nn.functional as F
```

```python
loss_function = F.mse_loss     #We are using the inbuilt loss function i.e. the mean square error
loss = loss_function(preds, targets)
opt = torch.optim.SGD(model.parameters(), lr = 1e-8)     #We now create a gradient descent optimizer and lr is the learning rate
```

```python
'''
This function takes the number of iterations(epochs), the type of neural network(model), the loss function used to determine
the accuracy(loss_fn), the optimzation factor(opt), and the training dataset(train_dl). After all the opertaions are done,
it returns the predicted values with minimum loss. If the results are not yet accurate, alter minor values like epoch and opt.
'''
def fit(epochs, model, loss_fn, opt, train_dl):
    for e in range(epochs):
        for xb, yb in train_dl:
            pred = model(xb)
            loss = loss_fn(pred, yb)
            loss.backward()
            opt.step()
            opt.zero_grad()
#         if e % 1000 == 0:
#             print(loss)
```

```python
fit(500, model, loss_function, opt, train_dl)
```
