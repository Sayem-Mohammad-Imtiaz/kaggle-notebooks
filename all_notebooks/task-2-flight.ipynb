```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import sklearn
import matplotlib.pyplot as plt
import math
df=pd.read_csv("../input/flight-take-off-data-jfk-airport/M1_final.csv")
print(df)
```

```python

df.drop('TAIL_NUM',axis=1,inplace=True)
df.shape
```

Lable Encoding

```python
from sklearn.preprocessing import LabelEncoder
l = LabelEncoder()
df.OP_UNIQUE_CARRIER=l.fit_transform(df.OP_UNIQUE_CARRIER)
df["DEST"]=l.fit_transform(df["DEST"])
df["Wind"] =l.fit_transform(df["Wind"].astype(str))
df["Condition"] =l.fit_transform(df["Condition"])
df['Dew Point'] = l.fit_transform(df['Dew Point'])
```

```python
X = df.drop(['TAXI_OUT'],axis = 1)
X.head()
df.shape
y = df.TAXI_OUT
y.shape
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.1)
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lr=LinearRegression()
lr=lr.fit(X_train,y_train)
y_actual = lr.predict(X_test)

lrerror = mean_squared_error(y_test,y_actual,squared = False)
lrerror
```

Ridge Regression

```python
from sklearn import linear_model
rid=linear_model.Ridge(alpha=1,max_iter=1000,tol=.01)
rid.fit(X_train,y_train)
pred1=rid.predict(X_test)
riderror=mean_squared_error(y_test,pred1,squared = False)
riderror
```

Lasso Regression

```python


from sklearn import linear_model
lareg=linear_model.Lasso(alpha=1,max_iter=1000,tol=.01)
lareg.fit(X_train,y_train)
pred2=lareg.predict(X_test)
laerror=mean_squared_error(y_test,pred2,squared = False)
laerror
```

K Nearest Neighbours

```python
from sklearn.neighbors import KNeighborsRegressor
kreg = KNeighborsRegressor(n_neighbors = 200)
kreg.fit(X_train,y_train)
pred3=kreg.predict(X_test)
kerror = mean_squared_error(y_test,pred3,squared = False)
kerror
```

SVR

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
sc1=StandardScaler()
sc2=StandardScaler()
x1=X_train.copy()
x2=X_test.copy()
x1=sc1.fit_transform(x1)
x2=sc2.fit_transform(x2)
svr=SVR()
svr.fit(x1,y_train)
pred4=svr.predict(x2)
svrerror=mean_squared_error(y_test,pred4,squared = False)
svrerror
```

Naive Bayes

```python

from sklearn.linear_model import BayesianRidge
breg=BayesianRidge()
breg.fit(X_train,y_train)
pred5=breg.predict(X_test)
berror=mean_squared_error(y_test,pred5,squared=False)
berror
```

Light GBM

```python
from lightgbm import LGBMRegressor
lgbmreg=LGBMRegressor()
lgbmreg.fit(X_train,y_train)
pred6=lgbmreg.predict(X_test)
lgbmerror=mean_squared_error(y_test,pred6,squared=False)
lgbmerror
```

Random Forests

```python
from sklearn.ensemble import RandomForestRegressor
rfreg = RandomForestRegressor(n_estimators=100)
rfreg.fit(X_train,y_train)
pred7 = rfreg.predict(X_test)
rferror = mean_squared_error(y_test,pred7,squared = False)
rferror
```

One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
p = (df.dtypes == 'object')
cols = list(p[p].index)
OH_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)
OH_cols_df = pd.DataFrame(OH_encoder.fit_transform(df[cols]))
```

```python
OH_cols_df.index = df.index
num_df = df.drop(cols, axis=1)
OH_df = pd.concat([num_df, OH_cols_df], axis=1)
OH_df.head()
y=OH_df.TAXI_OUT
X=OH_df.drop(['TAXI_OUT'], axis =1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.1)
```

Linear Regression

```python
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
loreg=LinearRegression()
loreg.fit(X_train,y_train)
predi=loreg.predict(X_test)
lroerror=mean_squared_error(y_test,predi,squared = False)
lroerror
```

Ridge Regression

```python


from sklearn import linear_model
rido=linear_model.Ridge(alpha=1,max_iter=1000,tol=.01)
rido.fit(X_train,y_train)
predi1=rido.predict(X_test)
ridoerror=mean_squared_error(y_test,predi1,squared = False)
ridoerror
```

Lasso Regression

```python
from sklearn import linear_model
laoreg=linear_model.Lasso(alpha=1,max_iter=1000,tol=.01)
laoreg.fit(X_train,y_train)
predi2=laoreg.predict(X_test)
laoerror=mean_squared_error(y_test,predi2,squared = False)
laoerror
```

K Nearest Neighbours Regression

```python
from sklearn.neighbors import KNeighborsRegressor
koreg = KNeighborsRegressor(n_neighbors = 200)
koreg.fit(X_train,y_train)
predi3=koreg.predict(X_test)
koerror = mean_squared_error(y_test,predi3,squared = False)
koerror
```

SVR

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
sc1=StandardScaler()
sc2=StandardScaler()
x1=X_train.copy()
x2=X_test.copy()
x1=sc1.fit_transform(x1)
x2=sc2.fit_transform(x2)
svro=SVR()
svro.fit(x1,y_train)
predi4=svro.predict(x2)
svroerror=mean_squared_error(y_test,predi4,squared = False)
svroerror

```

Naive Bayes

```python
from sklearn.linear_model import BayesianRidge
boreg=BayesianRidge()
boreg.fit(X_train,y_train)
predi5=boreg.predict(X_test)
boerror=mean_squared_error(y_test,predi5,squared=False)
boerror
```

Light GBM

```python
from lightgbm import LGBMRegressor
lgbmoreg=LGBMRegressor()
lgbmoreg.fit(X_train,y_train)
predi6=lgbmoreg.predict(X_test)
lgbmoerror=mean_squared_error(y_test,predi6,squared=False)
lgbmoerror
```

Random Forests

```python
from sklearn.ensemble import RandomForestRegressor
rforeg = RandomForestRegressor(n_estimators=100)
rforeg.fit(X_train,y_train)
predi7 = rforeg.predict(X_test)
rfoerror = mean_squared_error(y_test,predi7,squared = False)
rfoerror
```

Comparative Graphs.

```python
from matplotlib import style
error1=np.array([lrerror,riderror,laerror,kerror,svrerror,berror,lgbmerror,rferror])
error2=np.array([lroerror,ridoerror,laoerror,koerror,svroerror,boerror,lgbmoerror,rfoerror])
style.use("ggplot")
plt.plot(["lnrR","Ridge","Lasso","KNN","SVR","NB","LGBM","RF"],error1,"r",label="label encoding",linewidth=2)
plt.plot(["lnrR","Ridge","Lasso","KNN","SVR","NB","LGBM","RF"],error2,"b",label="one hot encoding",linewidth=2)
plt.title("Comparison")
plt.xlabel("models")
plt.ylabel("root mean squared error")
plt.legend() 
plt.grid(True,color="k")
plt.show
```
