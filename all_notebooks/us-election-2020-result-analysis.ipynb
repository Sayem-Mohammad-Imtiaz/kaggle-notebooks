# Load the dataset into a data frame using Pandas

```python
data = pd.read_csv("/kaggle/input/us-election-2020/president_county_candidate.csv")
data.head(5) 
```

```python
data.columns
```

# Top 5 candidates that got more votes

```python
max_voted_candidate = data.groupby('candidate')['votes'].sum()
max_voted_candidate = pd.DataFrame(max_voted_candidate)
max_voted_candidate = max_voted_candidate.sort_values(by= 'votes',ascending=False)
top5_max_voted_candidate = max_voted_candidate.head(5)
top5_max_voted_candidate = top5_max_voted_candidate.reset_index()
#top5_max_voted_candidate.keys()
top5_max_voted_candidate
```

```python
plt.figure(figsize = (12,5))
plt.bar(top5_max_voted_candidate['candidate'],top5_max_voted_candidate['votes'], color=[ 'green', 'red', 'blue', 'cyan', 'yellow'])
```

# Top Five Parties that got more votes

```python
max_voted_party = data.groupby('party')['votes'].sum()
max_voted_party = pd.DataFrame(max_voted_party)
top5_max_voted_party = max_voted_party.sort_values(by= 'votes',ascending=False)
top5_max_voted_party = top5_max_voted_party.head(5)
top5_max_voted_party = top5_max_voted_party.reset_index()
top5_max_voted_party
```

```python
plt.figure(figsize = (12,5))
plt.bar(top5_max_voted_party['party'],top5_max_voted_party['votes'], color=[ 'blue', 'red', 'green', 'cyan', 'yellow'])
```

# US Result Map Visualization

```python
president_state_data = pd.read_csv('/kaggle/input/us-election-2020/president_state.csv')
# datafile
state_codes = pd.read_csv('/kaggle/input/latitude-and-longitude-for-every-country-and-state/world_country_and_usa_states_latitude_and_longitude_values.csv')
# state_codes
datafile = president_state_data.merge(state_codes, left_on='state', right_on='usa_state')
```

```python
datafile.head(5)
```

```python
fig = px.choropleth(datafile,locations='usa_state_code', color="votes",
                           locationmode = 'USA-states',
                           scope="usa",
                           hover_name="state",
                           title='USA Presidential Votes Counts' 
                          )
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
