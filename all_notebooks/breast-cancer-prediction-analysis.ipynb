# Null value Handling

```python
for col in df.columns:
    if df[col].isnull().sum()!=0:
        print("The column "+ col + "has " + str(df[col].isnull().sum()) + " missing values")
df=df.drop(columns=['Unnamed: 32'])
```

No null values present in the dataset

# Discrete variables

```python
discrete_var=[feature for feature in df.columns if len(df[feature].unique()) <= 25 and feature not in ["id","diagnosis"]]
```

```python
discrete_var
```

# Continous variables

```python
continous_features=[feature for feature in df.columns if feature not in discrete_var+['id',"diagnosis"]]
```

```python
print(continous_features)
```

```python
for feature in continous_features:
    data=df.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.title(feature)
    plt.show()
```

# Outlier detection

```python
import seaborn as sns
for feature in continous_features:
    data=df.copy()
    if 0 in data[feature].unique():
        pass
    else:
        #data[feature]=np.log(data[feature])
        sns.boxplot(data[feature])
        #plt.xlabel(feature)
        plt.ylabel(feature)
        plt.title(feature)
        plt.show()
```

```python
area=['radius_mean','perimeter_mean','area_mean','area_worst','radius_worst','perimeter_worst']
df[area].corr()
```

As radius mean, perimeter mean, area_worst, radius_worst and perimeter_worst are highly correlated to area_mean we will use only area_mean

```python
#dropping the highly correlated features
droppable=[features for features in continous_features if "worst" in features]
```

```python
df["diagnosis"]=np.where(df['diagnosis']=='M',0,1)
df1=df.drop(columns=droppable+['radius_mean','perimeter_mean'])
```

```python
df1.head()
```

```python
df["diagnosis"].value_counts()
```

# scaling and splitting the train and test data

```python
X=df.iloc[:,2:]
X1=X=df1.iloc[:,2:]
```

```python
Y=df['diagnosis']
Y1=df1['diagnosis']
```

```python
X.shape
```

```python
Y.value_counts()
```

```python
SS=StandardScaler()
```

```python
X_scaled=SS.fit_transform(X)
X1_scaled=SS.fit_transform(X1)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X_scaled,Y,random_state=1,test_size=0.3)
X1_train,X1_test,Y1_train,Y1_test=train_test_split(X1_scaled,Y1,random_state=1,test_size=0.3)
```

# Model building (Logistic Regression)

```python
LR=LogisticRegression(solver = 'lbfgs')
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
LR.fit(X_train,Y_train)
Y_pred=LR.predict(X_test)
print(f1_score(Y_test,Y_pred))
print(accuracy_score(Y_test,Y_pred))
```

```python
LR1=LogisticRegression(solver = 'lbfgs')
LR1.fit(X1_train,Y1_train)
Y1_pred=LR1.predict(X1_test)
print(f1_score(Y1_test,Y1_pred))
print(accuracy_score(Y1_test,Y1_pred))
```
