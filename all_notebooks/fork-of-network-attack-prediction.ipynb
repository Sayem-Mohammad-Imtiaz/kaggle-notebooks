```python
num_arr = train_df.as_matrix("cont_"+ str(i) for i in range(1,19))
```

```python
num_arr.shape
```

```python
cat_arr = train_df.as_matrix("cat_"+ str(i) for i in range(1,24))
```

```python
cat_arr.shape
```

```python
target = train_df["target"].values
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import GaussianNB
```

```python
mul_nb = MultinomialNB()
mul_nb.fit(cat_arr,target)
```

```python
mul_nb.score(cat_arr,target)
```

```python
gau_nb = GaussianNB()
gau_nb.fit(num_arr,target)
```

```python
gau_nb.score(num_arr,target)
```

```python
X_train = mul_nb.predict_proba(cat_arr) * gau_nb.predict_proba(num_arr)
X_train.shape
```

```python
X_train
```

```python
Y = np.argmax(X_train,axis=1)
```

```python
def accuracy(Y,target):
    m = target.shape[0]
    acc = np.count_nonzero((Y-target)==0)/m
    return acc
```

```python
accuracy(Y,target)
```
