```python
### There is no function for calculating adjusted r2_score in SKLearn, so we need to do it our selves

def adj_r2(data_frame, r2, num_of_input_dimensions):
    #Adjusted RSquared of the model
    
    n=len(data_frame) #number of records 
    p=num_of_input_dimensions #number of features .i.e. columns excluding uniqueId and target variable
    return 1-(1-r2)*(n-1)/(n-p-1) # the mathematical formula for calculating adjusted r2
```

```python
num_dim=len(dataset.columns)-2
print(num_dim)
adjusted_r2=adj_r2(dataset,mul_r2,num_dim )
print(adjusted_r2)
```
