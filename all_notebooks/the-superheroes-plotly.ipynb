# **<h2 style="background-color:DodgerBlue;">CLEANING DATA</h2>**

<p style="background-color:Tomato;">The dataset contains lot of unnecessary columns ,missing values and irrelevant values which needs to be replaced and properly handled</p>

```python
#handling missing values
df.isnull().sum()
```


<p style="color:Violet;">The biography aliases columns have almost all null values and won't be used in analysis so we will drop them along with appearance race column</p>

```python
df.drop(["biography__aliases__002","biography__aliases__003","biography__aliases__004","biography__aliases__005",
        "biography__aliases__005","biography__aliases__006","biography__aliases__007","biography__aliases__008",
        "biography__aliases__009","biography__aliases__010","biography__aliases__011","biography__aliases__012",
        "biography__aliases__013","biography__aliases__014","biography__aliases__015","biography__aliases__016",
        "biography__aliases__017","biography__aliases__018","biography__aliases__019","biography__aliases__020",
        "appearance__race"
        ],axis= 1,inplace= True)
```

<p style="background-color:Tomato;">FILLING THE REST OF THE STATS COLUMNS WITH MEAN</p>

```python
df["powerstats__intelligence"]= df["powerstats__intelligence"].fillna(df["powerstats__intelligence"].mean())
df["powerstats__strength"]= df["powerstats__strength"].fillna(df["powerstats__strength"].mean())
df["powerstats__speed"]= df["powerstats__speed"].fillna(df["powerstats__speed"].mean())
df["powerstats__durability"]= df["powerstats__durability"].fillna(df["powerstats__durability"].mean())
df["powerstats__power"]= df["powerstats__power"].fillna(df["powerstats__power"].mean())
df["powerstats__combat"]= df["powerstats__combat"].fillna(df["powerstats__combat"].mean())
df["biography__publisher"]= df["biography__publisher"].fillna(df["biography__publisher"].mode())
```

<p style="background-color:Tomato;">DROPPING THE SECONDARY HEIGHT AND WEIGHT COLUMNS</p>

```python
df.drop(["appearance__height__002","appearance__weight__002"],axis=1,inplace= True)
```

```python
df.drop("biography__full-name",axis=1,inplace= True)
```

<p style="background-color:Tomato;">RENAMING COLUMNS TO MORE UNDERSTANDABLE FORMAT</p>

```python
df.rename(columns={"powerstats__intelligence":"intelligence","powerstats__strength":"strenth",
                   "powerstats__speed":"speed","powerstats__durability":"durability",   
                   "powerstats__power":"power","powerstats__combat":"combat",
                   "biography__alter-egos":"alter-ego","biography__aliases__001":"alliance",
                   "biography__place-of-birth":"origin","biography__first-appearance":"first-appearance",
                   "biography__alter-egos":"alter-ego","biography__aliases__001":"alias",
                   "biography__publisher":"publisher","biography__alignment":"alignment","appearance__gender":"gender",
                   "appearance__height__001":"height","appearance__weight__001":"weight","appearance__eye-color":"eye-color",
                   "appearance__hair-color":"hair-color","connections__group-affiliation":"group"
                  },inplace= True)
```

```python
df.rename(columns={"strenth":"strength"},inplace= True)
```

<p style="background-color:Tomato;">CLEANING HEIGHT AND WEIGHT COLUMN</p>

```python
df["height"].value_counts()
```

```python
df["height"].unique()
```

<p style="background-color:Tomato;">REPLACING IRRELEVANT VALUES FROM HEIGHT COLUMN</p>

```python
df["height"]= df["height"].apply(lambda x:str(x).replace("Shaker Heights, Ohio","6.0"))
```

```python
df["height"]= df["height"].apply(lambda x:str(x).replace("-","6.0"))
```

<p style="background-color:Tomato;">THE HEIGHT COLUMN VALUES HAVE " ' " AS A DELIMETER ,WE WILL REPLACE THE DELIMETER TO DECIMAL POINT SO THAT WE CAN CONVERT THE HEIGHT COLUMN TO NUMERIC FLOAT "-" </p>

```python
def main(x):
    str(x)
    if x[-1]== "'":
        return x[:-1]
    else:
        return x
```

```python
df["height"]=df["height"].apply(main)
```

```python
df["height"]=df["height"].apply(lambda x: str(x).replace("'","."))
df["height"]=df["height"].apply(lambda x: float(x))
```

<p style="background-color:Tomato;">REPLACING IRRELEVANT VALUES OF WEIGHT COLUMN AND ALSO REMOVING THE WEIGHT UNIT TO CONVERT THE COLUMN TO NUMERIC</p>

```python
df["weight"].value_counts()
```

```python
df["weight"]= df["weight"].apply(lambda x:str(x).replace("- lb","175 lb"))
```

```python
df["weight"]= df["weight"].apply(lambda x:str(x).replace(" lb",""))
df["weight"]= df["weight"].apply(lambda x:int(x))
```

```python
df.isnull().sum()
```

<p style="background-color:Tomato;">CLEANING THE GENDER,ALIGNMENT AND OTHER CATEGORICAL COLUMNS LIKE EYE AND HAIR</p>

```python
df.gender.value_counts()
```

```python
df.gender=df.gender.apply(lambda x:str(x).replace("-","unknown"))
```

```python
df.alignment=df.alignment.apply(lambda x:str(x).replace("-","neutral"))
```

```python
df["eye-color"].mode()
```

```python
df["eye-color"]= df["eye-color"].apply(lambda x:str(x).replace("-","Blue"))
```

```python
df["hair-color"].value_counts()
```

```python
df["hair-color"]= df["hair-color"].apply(lambda x:str(x).replace("-","Black"))
```

<p style="background-color:Tomato;">There are some duplicate characters in the dataset ,we will keep the first one and drop the rest</p>

```python
df.drop_duplicates(subset="name", keep="first", inplace=True)
```

<p style="background-color:Tomato;">ALL THE CLEANING PART IS DONE</p>
