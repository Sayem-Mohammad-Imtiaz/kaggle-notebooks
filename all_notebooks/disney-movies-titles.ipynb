# 'Disney'-shaped word cloud of film titles, sized by box office

```python
#necessary as the later use of io seems to overwrites the default path
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename=='walt_disney_movies.csv':
            path = os.path.join(dirname, filename)
        else:
            pass
```

```python
import numpy as np
import pandas as pd

from PIL import Image

from wordcloud import WordCloud,STOPWORDS
from matplotlib import pyplot as plt

import requests
import io
```

```python
ds = pd.read_csv(path)

bo = np.nan_to_num(ds["Box office"].values)
tt = ds.title.values
```

```python
wc = pd.Series(bo)
wc.index = tt
wd = wc[1:].to_dict()
```

```python
response = requests.get("https://raw.githubusercontent.com/ambader/ambader/main/MISC/disney_ll.png")
film_mask = np.array(Image.open(io.BytesIO(response.content)))

plt.figure(figsize= (30,15))
wordcloud = WordCloud(colormap="Blues",max_words=500,width = 500, height = 200, background_color ="aliceblue",min_font_size = 1, mask=film_mask, prefer_horizontal=0.5).generate_from_frequencies(wd)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
