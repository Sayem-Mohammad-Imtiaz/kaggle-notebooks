```python
dataset=pd.read_csv('../input/zomato-restaurants-in-delhi-ncr/DelhiNCR Restaurants.csv')
dataset.head()
```

```python
dataset['Locality_address']=dataset['Locality'].apply(lambda x: x.split(',')[0])
dataset.head()
```

```python
import matplotlib.pyplot as plt
plt.hist(dataset['Pricing_for_2'])
plt.ylabel('Frequency')
plt.xlabel('Price for 2')
plt.plot()
```

```python
unique_localities=dataset['Locality_address'].value_counts()
print('Number of unique localities are -',len(unique_localities))
```

```python
unique_loc=pd.DataFrame(unique_localities,index=None)
unique_loc=unique_loc.reset_index()
```

```python
unique_loc=unique_loc.rename(columns={'index':'Locality'})
unique_loc=unique_loc.rename(columns={'Locality_address':'No_of_eateries'})
```

```python
unique_loc_25=unique_loc.head(25)
```

```python
plt.figure(figsize=(15,12))
plt.bar(x=unique_loc_25['Locality'],height=unique_loc_25['No_of_eateries'])
plt.xlabel('Localities')
plt.xticks(rotation=90) 
plt.ylabel('No_of_eateries')
plt.title('25 of the most eatery dense areas')
plt.plot()
```

```python
#Average Rating for areas 
Rating_per_Area=pd.DataFrame(dataset['Dining_Rating'].groupby(dataset['Locality_address']).mean()).reset_index()
Rating_per_Area=Rating_per_Area.sort_values(by='Dining_Rating',ascending=False)
Rating_per_Area
```

```python

Rating_per_Area_25=Rating_per_Area.head(25)
plt.figure(figsize=(15,12))
plt.bar(x=Rating_per_Area_25['Locality_address'],height=Rating_per_Area_25['Dining_Rating'])
plt.xlabel('Localities')
plt.xticks(rotation=90) 
plt.ylabel('Average Ratings ')
plt.title('25 Highest Rated Areas')
plt.plot()
```

```python
Rating_per_Area_Bottom25=Rating_per_Area.tail(25)
plt.figure(figsize=(15,12))
plt.bar(x=Rating_per_Area_Bottom25['Locality_address'],height=Rating_per_Area_Bottom25['Dining_Rating'])
plt.xlabel('Localities')
plt.xticks(rotation=90) 
plt.ylabel('Average Ratings ')
plt.title('25 Lowest Rated Areas')
plt.plot()
```
