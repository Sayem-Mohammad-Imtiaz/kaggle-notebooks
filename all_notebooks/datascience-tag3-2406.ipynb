```python
data = pd.read_csv("/kaggle/input/cusersmarildownloadsearningcsv/earning.csv", delimiter=";")
data.info()
```

```python
data_red = data[["year", 
                 "femalesmanagers", 
                 "malemanagers", 
                 "femalemachineryoperatorsanddrivers", 
                 "malemachineryoperatorsanddrivers"]]
data_red.info()
```

```python
data_red.head()
```

```python
data_red.columns = ["year",
                   "manager_f",
                   "manager_m",
                   "driver_f",
                   "driver_m"]
data_red.head()
```

```python
plt.plot(data_red["year"], data_red["manager_m"] - data_red["manager_f"], label="manager")


plt.plot(data_red["year"], data_red["driver_m"] - data_red["driver_f"], label="driver")
plt.plot(data_red["year"], np.zeros(len(data_red["year"])), color = "red", linestyle="dotted")
plt.title("Lohnentwicklung m vs f")

plt.ylabel("Drchschn Stundenlohn in US $")
plt.xlabel("Jahr")
plt.legend()
plt.xlim(2004, 2017)


plt.style.use("seaborn")
plt.show()
```
