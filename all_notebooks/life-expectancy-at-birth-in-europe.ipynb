```python
data = pd.read_csv('/kaggle/input/life-expectancy-at-birth-in-europe/life_expectancy_at_birth.csv')
```

```python
data.drop(columns=['statinfo'], inplace=True)
print(list(data.columns))
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
data['sex'] = encoder.fit_transform(data['sex'])
data['geography'] = encoder.fit_transform(data['geography'])
print(data.head())
```

```python
data.isnull().sum()
```

```python
print(data[data['value'].isnull()])
```

```python
data['value'] = data['value'].fillna(data['value'].mode()[0]) # fill the null values with the highest frequency value of the list
```

```python
data.isnull().sum().sum()
```
