# Data

```python
tracks_data = pd.read_csv('/kaggle/input/spotify-dataset-19212020-160k-tracks/tracks.csv', )
```

```python
tracks_by_year = tracks_data.set_index('release_date').copy()
tracks_by_year.index = pd.to_datetime(tracks_by_year.index)
tracks_by_year.sort_index(inplace=True)
```

**I tried to find redundant data by checking for rows having same name or same audio features**

```python
audio_features = tracks_by_year.columns.tolist()
to_remove = ['id', 'artists', 'id_artists', 'popularity', 'duration_ms', 'explicit',
             'release_date', 'time_signature', 'key', 'mode']
reqd_cols = list(set(audio_features).difference(set(to_remove)))
reqd_cols
```

```python
duplicates = tracks_by_year[tracks_by_year.duplicated(subset=reqd_cols, keep=False)].copy()
```
