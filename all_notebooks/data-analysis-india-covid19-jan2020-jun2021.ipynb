```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/kaggle/input/covid19-data-of-india/covid_cases.csv')
```

```python
plt.figure(figsize =(100,4 ))
sns.relplot(x="Number of cases",y="Total Recovered",data=data,height=5, aspect=1.5)
plt.ticklabel_format(style = 'plain')
plt.show()
```

```python
sns.relplot(x="Number of cases",y="Daily Deceased",data=data,height=5, aspect=1.5)
plt.ticklabel_format(style = 'plain')
plt.show()
```

```python
sns.relplot(x="Timestamp",y="Daily Confirmed",data=data,hue="Daily Deceased",height=5, aspect=1.5)
plt.ticklabel_format(style='plain')
```

```python
sns.relplot(x="Timestamp",y="Daily Confirmed",data=data,hue="Daily Recovered",height=5, aspect=1.5)
plt.ticklabel_format(style='plain')
```

```python
sns.pairplot(data)
```

```python
plt.figure(figsize=(17,8))
sns.lineplot(x="Date",y="Daily Confirmed",data=data,color='yellow',label="Daily Confirmed")
sns.lineplot(x="Date",y="Daily Recovered",data=data,color='green',label="Daily Recovered")
sns.lineplot(x="Date",y="Daily Deceased",data=data,color='red',label="Daily Deceased")
plt.xlabel('Date: 30 Jan 2020 --> 03 Jun 2021')
plt.ylabel('Number of Confirmed/Recovered/Deceased')
plt.legend(loc="upper left")
```
