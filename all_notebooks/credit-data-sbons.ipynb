```python
# cleaning data
x = data.drop(['CUST_ID'], axis=1)
```

```python
### importing module for KMeans algorithm

from sklearn.cluster import KMeans

# Defining lists - cost and n_cluster in order to create elbow plot
cost = []
n_cluster = []

# using a for loop to make 9 models with different k values and store cost.
for k in range(1,10):
    kmeans_model = KMeans(n_clusters = k, random_state = 2)
    kmeans_model.fit(x)
    
    labels = kmeans_model.labels_
    
    inertia = kmeans_model.inertia_
    print("k = ",k ,"cost =",inertia)
    cost.append(inertia)
    n_cluster.append(k)
    
```

```python
import matplotlib.pyplot as plt

# Plotting elbow plot

plt.plot(n_cluster,cost)

plt.xlabel('number of cluster k')
plt.ylabel('cost')

plt.title('Elbow Plot')

plt.show()
```

```python
### Chose k as 2 using elbow plot 
# Redoing the KMeans model with k = 2

kmeans_model = KMeans(n_clusters = 2, random_state = 2)
kmeans_model.fit(x)

# Cluster label
labels = kmeans_model.labels_

```

```python
#Checking labels
labels
```

```python
# Creating cluster label data frame to add to data
labels_df = pd.DataFrame(labels)
```

```python
# adding column title to dataframe
labels_df.columns = ['cluster_id']
```

```python
# Checking dataframe
labels_df
```

```python
# Resetting index for existig data to match dataframe index
data = data.reset_index()
```

```python
#check
data
```

```python
# Dropping index column from data
data = data.drop(['index'],axis=1)
```

```python
#check
data
```

```python
# Concatenating dataframe and data
output = pd.concat([data,labels_df], axis =1)
```

```python
# Check output
output
```

```python
# Importing module to plot data clusters
import seaborn as sns
sns.scatterplot(data=output, x='BALANCE',y='CREDIT_LIMIT',hue="cluster_id")
```
