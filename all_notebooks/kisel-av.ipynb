```python
import numpy as np
import pandas as pd

train=pd.read_csv("/kaggle/input/covid-19-nlp-text-classification/Corona_NLP_train.csv",encoding='latin1')
test=pd.read_csv("/kaggle/input/covid-19-nlp-text-classification/Corona_NLP_test.csv",encoding='latin1')
```

Ввод данных
Дана таблица с сообщениями из твиттера. OriginalTweet - текст сообщения, а Sentiment - вручную проставленная, метка

Первые 5 тестовых элементов

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(train['OriginalTweet'])
print(X.shape)
```

```python
train['Sentiment'].head(20)
```

```python


senties = train['Sentiment']
a = list()
for i in senties:
    if (i == "Extremely Negative"):
        a.append("Negative")
    elif(i == "Extremely Positive"):
        a.append("Positive")
    else:
        a.append(i)
    
```

Т.к. изначально предлагалось 5 вариантов меток, с помощью данного "костыля" я сузил число вариантов до 3х

```python
#print(a)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression


X_train, X_test, y_train, y_test = train_test_split(X, a, test_size=0.01)
lgs = LogisticRegression(max_iter=1000000, solver='lbfgs')
lgs.fit(X_train, y_train)
print(lgs.score(X_test, y_test))
```

в среднем 80%-82% (логическая регрессия)

```python
X_predict = test['OriginalTweet']
X_predict = vectorizer.transform(X_predict)

y_Predict = lgs.predict(X_predict)
```

проверка модели на тестовых данных

```python
b = list()
for i in test['Sentiment']:
    if (i == "Extremely Negative"):
        b.append("Negative")
    elif(i == "Extremely Positive"):
        b.append("Positive")
    else:
        b.append(i)
```

тот же самый "костыль" чтобы было 3 возможных метки

```python
k=0
cool=0
for i in y_Predict:
    #print(i + "   vs   ")  
    #print(b[k])
    #print("===========")
    if (str(i) == str(b[k])):
        cool = cool + 1
    k = k + 1
    if (k==10000):
        break
print(cool/k)
```

точность в среднем 77%
