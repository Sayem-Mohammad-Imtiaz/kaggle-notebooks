# 1. Rank categories based on average enrollment

```python
fig, ax = plt.subplots(figsize=(12,7))
(df
 .groupby('Category')
 .apply(lambda x: x.Enrollment.mean())
 .sort_values(ascending=False)
).plot.bar(ax=ax, rot=0)
plt.ylabel('Enrollment', fontsize=16)
plt.xlabel('Category', fontsize=16)
plt.title('Average enrollment for each category', fontsize=20)
plt.show()
```

# 2. Find out which category has most courses


```python
fig, ax = plt.subplots(figsize=(16,10))
df.Category.value_counts().plot.pie(ax=ax, autopct='%1.2f%%', shadow=True, explode=(0.1, 0, 0, 0, 0), textprops={'size':'x-large', 'weight':'bold'})
plt.ylabel('')
plt.title('How many courses are in each category?', fontsize=20)
plt.show()
```

We see a disproportiante enrollment in tech courses although the number of tech courses on the Udemy platform is not disproportionate to the number of other courses. We can assume that people go on Udemy to mostly gain knowledge in technology.

# 4. How many courses have Python in their title? Machine Learning? Data Science?

```python
df.Title
```

```python
def python_in_title(title_list):
  title_count = []
  for i in title_list:
    if 'Python' in i:
      title_count.append(i)
  return len(title_count)

def machinelearning_in_title(title_list):
  title_count = []
  for i in title_list:
    if 'Machine Learning' in i:
      title_count.append(i)
  return len(title_count)

def datascience_in_title(title_list):
  title_count = []
  for i in title_list:
    if 'Data Science' in i:
      title_count.append(i)
  return len(title_count)
```

```python
print(python_in_title(list(df.Title)))
print(machinelearning_in_title(list(df.Title)))
print(datascience_in_title(list(df.Title)))
```

```python
fig, ax = plt.subplots(figsize=(12,8))
sns.barplot(x=['Python', 'Machine Leanring', 'Data Science'],y=[python_in_title(list(df.Title)), machinelearning_in_title(list(df.Title)), datascience_in_title(list(df.Title))])
plt.xlabel('Keyword in title', fontsize=16)
plt.ylabel('Amount of titles with the keyword', fontsize=16);
```
