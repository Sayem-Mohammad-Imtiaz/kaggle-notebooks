# Linear Regression using sklearn

```python
from sklearn.linear_model import LinearRegression 
```

```python
cars["transmission"] = cars["transmission"].astype('category').cat.codes
cars['fuelType'] = cars['fuelType'].astype('category').cat.codes
cars['model'] = cars['model'].astype('category').cat.codes
```

```python
features = [ 'tax','model' ,'year' , 'fuelType' , 'mileage','engineSize','mpg']
x = cars[features]
y = cars.price
```

```python
c = LinearRegression()
c.fit(x,y)
```

```python
print(c.intercept_)
print(c.coef_)
```

```python
print(list(zip(features,c.coef_)))
```

```python
c.predict([[100,100,100,100,100,100,100]]).round(2)
```

```python
c.score(x,y)
```

- inference:
1. stats model provides great summary.
2. R-squared: 78.7% which means our model can explain 78.7%   of our data
