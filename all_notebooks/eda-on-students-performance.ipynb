```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig, ax = plt.subplots(1,1, figsize = (10,7))
ax.bar(df['parental level of education'], df['math score'], width=0.8)
ax.set_xlabel('Parental level of eduction')
ax.set_ylabel('Maths score')
ax.set_title('Whose kids are having better math score?')
```

```python
fig, ax = plt.subplots(1,1, figsize = (10,7))
ax.scatter(df['math score'], df['reading score'], s=None,c=None, marker='*')
ax.set_xlabel('Maths score')
ax.set_ylabel('Reading Score')
ax.set_title('Math score Vs Reading score')
ax.legend(df.gender)
```
