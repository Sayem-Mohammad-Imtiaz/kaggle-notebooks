```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('../input/customer-segmentation-tutorial-in-python/Mall_Customers.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
def income_cat(inc):
    inc_cat='poor'
    
    if inc>=7 and inc<=80:
        inc_cat='rich'
    else:
        inc_cat='very rich'
    return inc_cat
def age_cate(age):
    age_cat='young'
    if age<=25:
        age_cat='young'
    elif age>25 and age<40:
        age_cat='elder'
    elif age>=40 and age<60:
        age_cat='old'
    else:
        age_cat='very old'
    return age_cat
def sc_cat(sco):
    sco_cat='very low'
    if sco<20:
        sco_cat='very low'
    elif sco>=20 and sco<40:
        sco_cat='low'
    elif sco>=40 and sco<60:
        sco_cat='moderate'
    elif sco>=60 and sco>80:
        sco_cat='high'
    else:
        sco_cat='very high'
    return sco_cat
```

```python
df['age_cat']=df['Age'].apply(age_cate)
df['income_cat']=df['Annual Income (k$)'].apply(income_cat)
df['score_cat']=df['Spending Score (1-100)'].apply(sc_cat)
```
