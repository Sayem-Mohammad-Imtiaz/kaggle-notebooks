This is a simple exploration of the data from this perspective.  I hope in the future to start building predictive models, but I just want to start simple

```python
df =pd.read_csv('../input/train.csv')
df.shape
```

#Kickstarter goal are distributed on a logarithmic scale
to observe a normal distribution of goals we can do a log-transform

```python
np.log10(df['goal']).plot.hist(bins = 35)
plt.title('Kickstarter Goals are on a Logarithmic Scale')
plt.xlabel('log10 Goal')
```

#$1million is the threshold of which people are willing to contribute 

There are only 4 projects funded above that amount projects 

```python
df['loggoal'] = np.log10(df['goal'])
sns.lmplot(x = 'loggoal', y = 'backers_count', col ='final_status', data = df, fit_reg = False)
```

#If you set your goal above $1 Million, people won't even bother backing you

```python
sns.lmplot(x = 'loggoal', y = 'backers_count', data = df[df['final_status'] == 0], fit_reg = False)
plt.title('Unsuccesful Kickstarters rarely find backers with goals above $1 Million')
```

```python
df[df['goal']>1000000].shape # in fact only 232 projects had goals over $1million
```

```python
df[df['goal']>1000000]['final_status'].value_counts() # and only 4 of those projects got funded
```

```python
plt.figure(figsize = (6,6))
sns.boxplot(x ='final_status', y = 'loggoal', data = df)
plt.title('Successful Kickstarters have on average lower Goals')
```

```python
df.groupby('final_status')['goal'].median().T
```

##the median successful kickstarter had a goal of $3,050
## while the median unsuccesful kickstarter had a goal of $5,400
###The Lesson thus far is to dream appropriately, but not over $1 Million
