```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

Dataset = pd.read_csv('/kaggle/input/chess/games.csv')

print("Objectif :")
print("Cette étude vise à analyser la différence de résultats entre les blancs et les noirs aux échecs.")
print("\n")

############Calcul du nombre de parties
print("Etude de l'échantillon :")
NbGames = Dataset.shape[0]
NbRatedGame = Dataset.loc[Dataset['rated']==1,:].shape[0]
print("Nombre de parties : " + str(NbGames) +" dont " + str(round(NbRatedGame/NbGames,3)*100) + "% classes")
#Calcul du nombre de joueurs
NbPlayer = Dataset['white_id'].append(Dataset['black_id']).nunique()
print("Nombre de joueurs : " + str(NbPlayer))
#calcul du classement
ListRate =  Dataset['white_rating'].append(Dataset['black_rating'])
print("La moyenne de classement des joueurs dans ses parties est de : " + str(ListRate.mean()))
print("Le classement max est de : " + str(ListRate.max()))
print("Le classement min est de : " + str(ListRate.min()))
print("\n")
print("Etude des victoires :")
#Calcul du pourcentage de victoire des blancs, des noirs et du pourcentage de match nul
NbWhiteWInner = Dataset.loc[Dataset['winner']=='white',:].shape[0]
NbBlackWInner = Dataset.loc[Dataset['winner']=='black',:].shape[0]
NbDraw = Dataset.loc[Dataset['winner']=='draw',:].shape[0]
print("Nombre de victoire des blancs : " + str(NbWhiteWInner) + " soit  " + str(round(NbWhiteWInner/NbGames,3)*100) +"% des parties totales")
print("Nombre de victoire des noirs : " + str(NbBlackWInner) + " soit " + str(round(NbBlackWInner/NbGames,3)*100) +"% des parties totales")
print("Nombre de match nul : " + str(NbDraw) + " soit " + str(round(NbDraw/NbGames,3)*100) +"% des parties totales")
print("\n")
#Pourcentage des types de victoires ( mate, draw, resign, outoftime )
NbResign = Dataset.loc[Dataset['victory_status']=='resign',:].shape[0]
NbMate = Dataset.loc[Dataset['victory_status']=='mate',:].shape[0]
NbOutOfTime = Dataset.loc[Dataset['victory_status']=='outoftime',:].shape[0]
NbPatStatus = Dataset.loc[Dataset['victory_status']=='draw',:].shape[0]

print("Nombre de victoire par abandon : " + str(NbResign) + " soit  " + str(round(NbResign/NbGames,3)*100) +"% des parties totales")
print("Nombre de victoire par mate : " + str(NbMate) + " soit  " + str(round(NbMate/NbGames,3)*100) +"% des parties totales")
print("Nombre de victoire par temps écoulé : " + str(NbOutOfTime) + " soit  " + str(round(NbOutOfTime/NbGames,3)*100) +"% des parties totales")
print("Nombre de pat : " + str(NbPatStatus) + " soit " + str(round(NbPatStatus/NbGames,3)*100) +"% des parties totales")
print("\n")

#Croisement des donnees de victoire et de couleur
print(pd.crosstab(Dataset['winner'], Dataset['victory_status'],normalize='index'))
print("On constate que les blancs ou noirs ont les mêmes statistiques de type de victoires. La majorité des parties se soldent par un abandon, un tiers des partie par un mat et moins d'un dixi�me des parties par manque de temps ")

print ("\n")
#Calcul du nombre du pourcentage de victoire de joueur mieux classé en blanc puis en noir
nbStrongWhite = Dataset.query('white_rating > black_rating and winner == "white"', inplace = False).shape[0]
nbStrongBlack = Dataset.query('white_rating < black_rating and winner == "black"', inplace = False).shape[0]
print("Pourcentage de victoire des blancs avec un classement supérieur : " + str(round(nbStrongWhite/NbWhiteWInner,3)*100) + "%")
print("Pourcentage de victoire des noirs avec un classement supérieur : " + str(round(nbStrongBlack/NbBlackWInner,3)*100) + "%")
print ("\nObservations et conlusion\n")
print("Lors de cette étude on constate donc que les blancs démarre légèrement favori en règle général (4% de probabilité en plus de gagner). \nLe moyen de gagner des blancs quant à lui n'est en moyenne pas très différent de celui des noirs. \nOn constate seulement 1% de victoire en plus par échec et mate tandis que les victoires sur le temps ou les abandons sont légèrement inférieurs aux types de victoires noirs.\nEn cas de niveau inférieur les blancs seront une fois de plus légèrement avantagé par rapport au noir, ce qui reste cohérent avec le taux de victoire légèrement plus élevé")
```
