```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_set = pd.read_csv('/kaggle/input/cholera-dataset/data.csv')
data_set.head()
```

```python
print('--------- Shape ---------')
print(data_set.shape)
print('--------- No. of Contries ---------')
print(data_set['Country'].nunique())
print('--------- Datatype of Columns ---------')
print(data_set.dtypes)
print('--------- Null Items ---------')
print(data_set.isnull().sum())
```

Now we have 2 problems to solve before playing with data
1. Datatype of Number of reported cases of cholera, Number of reported deaths from cholera, Cholera case fatality rate is Object
2. Null Cells

```python
def check_int(s):
    try: 
        float(s)
        return True
    except ValueError:
        return False
```

```python
def print_wrong_nos(arr):
    for i in range (len(arr)):
        if check_int(arr[i]) == False:
            print(i,arr[i])
```

```python
number_of_cases = data_set['Number of reported cases of cholera']
print_wrong_nos(number_of_cases)
```

```python
number_of_deaths = data_set['Number of reported deaths from cholera']
print_wrong_nos(number_of_deaths)
```

```python
fatality_rate = data_set['Cholera case fatality rate']
print_wrong_nos(fatality_rate)
```

```python
print(data_set.iloc[761].values)
print(data_set.iloc[1059].values)
```

```python
country_grp = data_set.groupby('Country')
country_grp.get_group('Iraq')
```

```python
data_set = data_set.drop([1059],axis = 0)
data_set.shape
```

```python
country_grp.get_group('Germany')
```

```python
data_set.iloc[761,3:5] = 0
data_set.iloc[761]
```

```python
data_set = data_set.fillna(0)
data_set.isnull().sum()
```

```python
##### correct data types
data_set['Country'] = data_set.Country.astype("string")
data_set['Year'] = data_set['Year'].astype("int")
data_set['Number of reported cases of cholera'] = data_set['Number of reported cases of cholera'].astype("int64")
data_set['Number of reported deaths from cholera'] = data_set['Number of reported deaths from cholera'].astype("int64")
data_set['Cholera case fatality rate'] = data_set['Cholera case fatality rate'].astype("float")
data_set['WHO Region'] = data_set['WHO Region'].astype("string")
data_set.dtypes
```

```python
data_set.describe()
```

```python
country_wise_cases = data_set.groupby('Country')['Number of reported cases of cholera'].sum()
major_countries = country_wise_cases[country_wise_cases > 0.1e6]
plt.figure(figsize=(20,7))
sns.barplot(x = major_countries.index,y = major_countries)
```

```python
df = pd.DataFrame(data={'Total Cases':major_countries})
sns.heatmap(data=df, annot=True)
```

```python
country_wise_deaths = data_set.groupby('Country')['Number of reported deaths from cholera'].sum()
major_countries = country_wise_deaths[country_wise_deaths>5000]
plt.figure(figsize=(20,7))
sns.barplot(x = major_countries.index,y = major_countries)
```

```python
df = pd.DataFrame(data={'Total Deaths':major_countries})
sns.heatmap(data=df, annot=True)
```
