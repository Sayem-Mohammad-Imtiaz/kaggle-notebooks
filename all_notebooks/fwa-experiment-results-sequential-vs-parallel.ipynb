```python
columns = ['size', 'sequential', 'parallel']
dense = pd.read_csv('../input/fwaresults/dense_fwa_sequential_and_parallel_results.csv', header=None, names = columns, skiprows=1)
sparse = pd.read_csv('../input/fwaresults/sparse_fwa_sequential_and_parallel_results.csv', header=None, names = columns, skiprows=1)

for i in dense.index:
    dense['sequential'][i] = float(dense['sequential'][i][:-2])
    dense['parallel'][i] = float(dense['parallel'][i][:-2])

for i in sparse.index:
    sparse['sequential'][i] = float(sparse['sequential'][i][:-2])
    sparse['parallel'][i] = float(sparse['parallel'][i][:-2])
    

display(dense)
display(sparse)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

ax = dense.plot.line(x="size", marker="o");
ax.set_ylabel("time");

bx = sparse.plot.line(x="size", marker="o");
bx.set_ylabel("time");
```
