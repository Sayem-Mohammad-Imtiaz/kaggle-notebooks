```python
import pandas as pd
import numba as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
data=pd.read_csv("../input/nifty50-data-20162021/NIFTY-50 data -2016-2021.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data['Date']=pd.to_datetime(data['Date'])
```

```python
data['Day']=data['Date'].dt.day_name()
```

```python
data.head()
```

```python
data.sort_values(by="Open")
```

```python
plt.figure(figsize=(12,7))
plt.style.use('fivethirtyeight')
plt.plot(data['Date'],data['Open'],label='Open',linestyle="--")
plt.plot(data['Date'],data['Close'],label='Close')
plt.tight_layout()
plt.legend()
```

```python
plt.figure(figsize=(12,7))
plt.style.use('fivethirtyeight')
plt.plot(data['Date'],data['High'],label='High')
plt.plot(data['Date'],data['Low'],label='Low')
plt.tight_layout()
plt.legend()
```
