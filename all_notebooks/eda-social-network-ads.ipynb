<h3 style = "font-family: Comic Sans MS;color:violet">To manage the data I am going to perform an feature scaling technique which can transform the original values into specific range</h3>

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
minmax = MinMaxScaler(feature_range=(1,5))
```

```python
features = ["Age","EstimatedSalary"]
```

```python
cols = df[features]
```

```python
df[features] = minmax.fit_transform(cols)
```
