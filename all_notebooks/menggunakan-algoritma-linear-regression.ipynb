Setelah kita mengetahui karakteristik dari setiap data, selanjutnya adalah tahap modeling

**Modeling**

```python
#1, membuat variabel x dan y
x = df.drop(columns='harga')
y = df['harga']

#2, split data menjadi training and testing dengan porsi 80:20.
x_train, x_test, y_train, y_test = train_test_split(x, y,train_size = 0.8,
                                                    test_size = 0.2, random_state=10)

#3, membuatobject linear regresi.
lin_reg = LinearRegression()

#4, train the model menggunakan training data yang sudah displit.
lin_reg.fit(x_train, y_train)
```

**Melihat Scre Akurasi**

```python
lin_reg.score(x_test, y_test)
```

**Melakukan Uji coba prediksi**

```python
lin_reg.predict([[8000,4,2]])
```
