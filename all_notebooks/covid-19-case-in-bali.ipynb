```python

import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

Reading the Data
*Data only contains cases around 2020*

```python
data = pd.read_csv('../input/covid19-case-in-bali/covid-19 case in Bali.csv')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

Get the aggregate of positive (positif) cases in every province

```python
total_kasus_p=data.groupby('provinsi', as_index=False).agg({"positif": "sum"})
total_kasus_p
```

The visualitation of positive cases in every province with bar graph

```python
dff = total_kasus_p.sort_values(by='positif', ascending=False)
fig = px.bar(total_kasus_p, x='provinsi', y='positif', height=400, color='positif',
             title='Total Kasus Positif 2020',
             text='positif',
             color_discrete_sequence= px.colors.qualitative.T10)
fig.update_xaxes(showgrid=False)
fig.update_yaxes(categoryorder='total ascending', ticksuffix='  ')
fig.update_traces(hovertemplate=None, marker=dict(line=dict(width=0)))
fig.update_layout(margin=dict(t=110, b=50, l=70, r=40),
                  xaxis_title=' ', yaxis_title=" ",
                  hovermode="y unified",
                  plot_bgcolor='#2d3035', paper_bgcolor='#2d3035',
                  title_font=dict(size=25, color='#a5a7ab',
                                  family="Lato, sans-serif"),
                  font=dict(color='#8a8d93'))
```

Get the agregate of positive,recovery and death cases in every province

```python
total_positif=data.groupby('provinsi', as_index=False).agg({"positif": "sum"})
total_positif
```

```python
total_sembuh=data.groupby('provinsi', as_index=False).agg({"sembuh": "sum"})
total_sembuh
```

```python
total_meninggal=data.groupby('provinsi', as_index=False).agg({"meninggal": "sum"})
total_meninggal
```

The visualitation of positive,recovery and death cases in every province

```python


trace1 = go.Bar(x=total_positif["provinsi"], y=total_positif["positif"],
                name="Positif", marker=dict(color="#DB6574"))
trace2 = go.Bar(x=total_sembuh["provinsi"], y=total_sembuh["sembuh"],
                name="Sembuh", marker=dict(color="#03DAC5"))
trace3 = go.Bar(x=total_meninggal["provinsi"], y=total_meninggal["meninggal"],
                name="Meninggal", marker=dict(color="blue"))
data_trace = [trace1, trace2, trace3]
fig_line = go.Figure(data_trace)

fig_line.update_traces(hovertemplate=None, marker=dict(line=dict(width=0)))
fig_line.update_xaxes(showgrid=False)
fig_line.update_yaxes(showgrid=False)
fig_line.update_layout(title='Kasus Positif,Sembuh,Meninggal',
                       margin=dict(t=100, b=0, l=70, r=40), height=350, 
                       hovermode="x unified", barmode='stack',
                       xaxis_title=' ', yaxis_title=" ",
                       plot_bgcolor='#333', paper_bgcolor='#333',
                       title_font=dict(size=25, color='#8a8d93', family="Lato, sans-serif"),
                       font=dict(color='#8a8d93'),
                       legend=dict(orientation="h", yanchor="bottom", y=1, xanchor="center", x=0.5))

```

Get the 3 top most cases and show the graph in line

```python
data_denpasar=data[data.provinsi=='denpasar']
data_denpasar
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=data_denpasar["bulan"],
    y=data_denpasar["positif"],
    name="Positif"
))

fig.add_trace(go.Scatter(
    x=data_denpasar["bulan"],
    y=data_denpasar["sembuh"],
    name="Sembuh"
))

fig.add_trace(go.Scatter(
    x=data_denpasar["bulan"],
    y=data_denpasar["meninggal"],
    name="Meninggal"
))

fig.add_trace(go.Scatter(
    x=data_denpasar["bulan"],
    y=data_denpasar["aktif"],
    name="Aktif"
))

fig.update_layout(legend_title_text='Kasus')
fig.show()
```
