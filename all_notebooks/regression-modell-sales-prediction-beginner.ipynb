```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
```

Lets import the Data

```python
df= pd.read_csv("/kaggle/input/dummy-advertising-and-sales-data/Dummy Data HSS.csv")
```

```python
df.info()
df.head()
```

Is there missing data in the dataset

```python
df.isnull().sum()
```

We have to add the missing data or delete it

```python
df = df.fillna(df.mean())
```

```python
df.isnull().sum()
```

# Data Visualisation

```python
sns.scatterplot(data=df,x="TV",y="Sales")
```

```python
sns.scatterplot(data=df,x="Radio",y="Sales")
```

```python
sns.scatterplot(data=df,x="Social Media",y="Sales")
```

# Preperation for the Modelling

```python
df.Influencer = df.Influencer.replace(["Nano","Micro","Macro","Mega"],[1,2,3,4])
```

# Linear Regression

```python
X = df.drop("Sales",axis="columns")
y= df.Sales
train_X,valid_X,train_y,valid_y = train_test_split(X,y,test_size=0.2)
lm = LinearRegression()
lm.fit(train_X,train_y)
print(lm.intercept_)
print(lm.coef_)
lm.score(train_X,train_y)
```

R^2 is good

```python
lm.fit(valid_X,valid_y)
sns.barplot(valid_X.columns,lm.coef_)
lm.score(valid_X,valid_y)
print(lm.intercept_)
print(lm.coef_)
```

The variable "TV" has the most impact on the sales
