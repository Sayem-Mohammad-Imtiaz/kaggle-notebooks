```python
def find_codes():
    venue1=str(input('Enter venue = '))
    inn1=str(input('Enter inn1 team='))
    inn2=str(input('Enter inn2 team='))
    for i in venue_codes['venue']:
        if i==venue1:
            print('code for',i,'is= ',venue_codes[venue_codes['venue']==i])
            break
    for k in inn1_codes['inn1']:
        if k==inn1:
            print('code for',k,'is= ',inn1_codes['code'])
            break
    for j in inn2_codes['inn2']:
        if j==inn2:
            print('code for',j,'is= ',inn2_codes['code'])
            break
```
