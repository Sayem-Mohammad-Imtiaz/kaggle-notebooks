```python
train_filepath = "../input/non-creditworthiness/credit-data-training.csv"
test_filepath = "../input/non-creditworthiness/customers-to-score.csv"
x = pd.read_csv(train_filepath)
x_test= pd.read_csv(test_filepath)

y = x['Credit-Application-Result']
```

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
feature = ['Duration-of-Credit-Month', 'Credit-Amount','Instalment-per-cent', 'Most-valuable-available-asset',
           'Age-years', 'Type-of-apartment']
x = x.fillna(35)
X = x[feature].copy()
X_test = x_test[feature].copy()

X_train, X_valid ,y_train, y_valid = train_test_split(X,y,train_size=0.8, test_size=0.2,random_state=0) 
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=1000, random_state=0)
classifier.fit(X_train, y_train) 
preds = classifier.predict(X_test)
count = (preds == 'Creditworthy')

count.sum()
```

```python
output = pd.DataFrame({'ID' : X_test.index, 'Age' : X_test['Age-years'], 'Duration-of-Credit-Month' : X_test['Duration-of-Credit-Month'],
                       'Credit': preds})
output.to_csv('submission.csv', index=False)
```
