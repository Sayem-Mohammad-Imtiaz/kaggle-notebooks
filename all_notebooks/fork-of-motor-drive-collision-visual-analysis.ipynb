```python
data = pd.read_csv("../input/nypd-motor-vehicle-collisions.csv")
data.head(10)
```

```python
V_nb = (data.loc[:,"BOROUGH"].dropna())
V_nbgh = np.unique(V_nb)
print (V_nbgh)
```

```python
#Quantity of people died per hood/ Quantity of accidents per hood = Mortality Ratio per hood
V_Killed_Qty = data[data["NUMBER OF PERSONS KILLED"] >0]
V_Killed_Qty.head(10)
```

```python
V_Kill_Hood = list()
for i in V_nbgh:
    Aux_Qty = V_Killed_Qty[V_Killed_Qty["BOROUGH"]==i]
    Aux_Kill = sum(Aux_Qty.loc[:,"NUMBER OF PERSONS KILLED"])
    Aux_Mort = Aux_Kill/len(Aux_Qty)
    V_Kill_Hood.append(Aux_Mort)
print (V_Kill_Hood)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.bar(V_nbgh,V_Kill_Hood)
plt.title("Mortality Ratio per borough")
plt.show()
```
