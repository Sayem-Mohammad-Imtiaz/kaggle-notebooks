```python
import plotly.express as px
import pandas as pd 
from pandas_profiling import ProfileReport
data=pd.read_csv("/kaggle/input/bollywood-box-office-20172020/bollywood_box_clean.csv")
data.head()
```

```python
profile=ProfileReport(data,title="Pandas Profiling Report")
```

```python
profile.to_notebook_iframe()
```

# Top 10 Movies With Highest Collections Worldwide

```python
high_earning=data.sort_values('movie_total_worldwide', ascending = False)
high_earning=high_earning.head(10)
px.bar(high_earning, y="movie_total_worldwide",x="movie_name",title="Top 10 Highest Box Office Collections ",color="movie_name")
```

# movies with Least Collection on Box office <br>statrting least 10 movies having value 0


```python
low_earning=data.sort_values('movie_total_worldwide',ascending=True)
low_earning=low_earning.head(50)
px.bar(low_earning,x="movie_name",y="movie_total_worldwide",title="Movies With Least Collections",color="movie_name")
```

```python
movie_open=data.sort_values('movie_opening',ascending=False)
movie_open=movie_open.head(10)
px.bar(movie_open,y='movie_opening',x='movie_name',title='Movies With Top 10 First Day Collections',color="movie_name")
```

```python
topgenre = data.groupby(["movie_genre"])["movie_total_worldwide"].sum().reset_index()
px.scatter(topgenre,'movie_genre','movie_total_worldwide',size="movie_total_worldwide",color="movie_genre",title="Which Genre is most liked by Audience")
```

```python
topdir = data.groupby(["movie_director"])["movie_total_worldwide"].sum().reset_index()
px.bar(topdir,"movie_director","movie_total_worldwide",title="Directors & their Movie Earnings")
```

```python
topdir=topdir.sort_values("movie_total_worldwide",ascending=False)
topdir=topdir.head(10)
px.bar(topdir,"movie_director","movie_total_worldwide",title="Top 10 Directors With Highest Movie Earnings ",color="movie_director")
```
