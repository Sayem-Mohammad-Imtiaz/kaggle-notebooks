# Bank Customer Churning Prediction

### 1 Problem Statement

A manager at the bank is disturbed with more and more customers leaving their credit card services. They would really appreciate if one could predict for them who is gonna get churned so they can proactively go to the customer to provide them better services and turn customers' decisions in the opposite direction.

#### Questions answered in the project:
1. Predict the features / characteristics of customers that would lead to them leaving the bank as customers (Main).
2. Other insights that would lead to solutions for better services of those customers about to leave.

### 2 Understanding Dataset

```python
df = pd.read_csv("../input/credit-card-customers/BankChurners.csv")
df.head(20)
```

```python
df.drop(columns=['Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1',
                'Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2'], inplace = True)
df.head(20)
```

We see the following columns: <br />
CLIENTNUM: Client number. Unique identifier for the customer holding the account <br />
Attrition_Flag: Internal event (customer activity) variable - if the account is closed then 1 else 0 <br />
Customer_Age: Demographic variable - Customer's Age in Years <br />
Gender: Demographic variable - M=Male, F=Female <br />
Dependent_count: Demographic variable - Number of dependents <br />
Education_Level: Demographic variable - Educational Qualification of the account holder (example: high school, college graduate, etc.) <br />
Marital_Status: Demographic variable - Married, Single, Divorced, Unknown <br />
Income_Category: Demographic variable - Annual Income Category of the account holder (< $40K, $40K - 60K, $60K - $80K, $80K-$120K, >120k, Unknown) <br />
Card_Category: Product Variable - Type of Card (Blue, Silver, Gold, Platinum) <br />
Months_on_book: Period of relationship with bank <br />
Total_Relationship_Count: Total no. of products held by the customer <br />
Months_Inactive_12_mon: No. of months inactive in the last 12 months <br />
Contacts_Count_12_mon: No. of Contacts in the last 12 months <br />
Credit_Limit: Credit Limit on the Credit Card <br />
Total_Revolving_Bal: Total Revolving Balance on the Credit Card (Unpaid balance) <br />
Avg_Open_To_Buy: Open to Buy Credit Line (Average of last 12 months) (Average difference between the credit limit and the present balance on the account) <br />
Total_Amt_Chng_Q4_Q1: Change in Transaction Amount (Q4 over Q1)<br />
Total_Trans_Amt: Total Transaction Amount (Last 12 months) <br />
Total_Trans_Ct: Total Transaction Count (Last 12 months) <br />
Total_Ct_Chng_Q4_Q1: Change in Transaction Count (Q4 over Q1) <br />
Avg_Utilization_Ratio: Average Card Utilization Ratio <br />

#### 3.2 Attrition_Flag
If the account is closed then 1 else 0 (output)



```python
print (sorted(df['Attrition_Flag'].unique()))
```

```python
''' 
Categorise Attrition_Flag
0: Existing Customer
1: Attrited Customer
'''

def flag(x):
    if x['Attrition_Flag'] == 'Existing Customer':
        return 0
    elif x['Attrition_Flag'] == 'Attrited Customer':
        return 1

df['Attrition_Flag'] = df.apply(flag, axis = 1)
```

##### 3.3 Customer_Age 

```python
map3 = sns.distplot(df.Customer_Age)
```

```python
print (sorted(df['Customer_Age'].unique()))
```

```python
''' 
Categorise age into groups of 5:
0: 26 - 30 
1: 31 - 35
2: 36 - 40
3: 41 - 45
4: 46 - 50
5: 51 - 55
6: 56 - 60
7: > 60
'''

def age(x):
    if x['Customer_Age'] >= 26 and x['Customer_Age'] <= 30:
        return 0
    elif x['Customer_Age'] >= 31 and x['Customer_Age'] <= 35:
        return 1
    elif x['Customer_Age'] >= 36 and x['Customer_Age'] <= 40:
        return 2
    elif x['Customer_Age'] >= 41 and x['Customer_Age'] <= 45:
        return 3
    elif x['Customer_Age'] >= 46 and x['Customer_Age'] <= 50:
        return 4
    elif x['Customer_Age'] >= 51 and x['Customer_Age'] <= 55:
        return 5
    elif x['Customer_Age'] >= 56 and x['Customer_Age'] <= 60:
        return 6
    else: #> 60: most likely retired
        return 7

df['Customer_Age'] = df.apply(age, axis = 1)
    
```

##### 3.4 Gender

```python
map4 = sns.catplot(x = 'Gender', kind="count", data = df)
```

```python
''' 
Categorise Gender
0: F
1: M
'''

def gender(x):
    if x['Gender'] == 'F':
        return 0
    elif x['Gender'] == 'M':
        return 1

df['Gender'] = df.apply(gender, axis = 1)
```

#### 3.5 Education_Level

```python
print (sorted(df['Education_Level'].unique()))
```

```python
(df['Education_Level'] == 'Unknown').sum()
```

```python
map5 = sns.catplot(x = 'Education_Level', kind="count", data = df, aspect = 2)
```

```python
''' 
Categorise Education_Level
0: Unknown 
1: Uneducated
2: High School
3: College
4: Graduate
5: Post-Graduate
6: Doctorate
'''

def edulvl(x):
    if x['Education_Level'] == 'Unknown':
        return 0
    elif x['Education_Level'] == 'Uneducated':
        return 1
    elif x['Education_Level'] == 'High School':
        return 2
    elif x['Education_Level'] == 'College':
        return 3
    elif x['Education_Level'] == 'Graduate':
        return 4
    elif x['Education_Level'] == 'Post-Graduate':
        return 5
    elif x['Education_Level'] == 'Doctorate':
        return 6

df['Education_Level'] = df.apply(edulvl, axis = 1)
    
```

#### 3.6 Marital_Status

```python
print (sorted(df['Marital_Status'].unique()))
```

```python
map6 = sns.catplot(x = 'Marital_Status', kind="count", data = df, aspect = 2)
```

```python
''' 
Categorise Marital_Status
0: Unknown 
1: Single
2: Married
3: Divorced
'''

def marstatus(x):
    if x['Marital_Status'] == 'Unknown':
        return 0
    elif x['Marital_Status'] == 'Single':
        return 1
    elif x['Marital_Status'] == 'Married':
        return 2
    elif x['Marital_Status'] == 'Divorced':
        return 3

df['Marital_Status'] = df.apply(marstatus, axis = 1)
    
```

#### 3.7 Income_Category

```python
print (sorted(df['Income_Category'].unique()))
```

```python
map7 = sns.catplot(x = 'Income_Category', kind="count", data = df, aspect = 2)
```

```python
''' 
Categorise Income_Category
0: Unknown 
1: Less than $40K
2: $40K - $60K
3: $60K - $80K
4: $80K - $120K
5: $120K +
'''

def income(x):
    if x['Income_Category'] == 'Unknown':
        return 0
    elif x['Income_Category'] == 'Less than $40K':
        return 1
    elif x['Income_Category'] == '$40K - $60K':
        return 2
    elif x['Income_Category'] == '$60K - $80K':
        return 3
    elif x['Income_Category'] == '$80K - $120K':
        return 4
    elif x['Income_Category'] == '$120K +':
        return 5

df['Income_Category'] = df.apply(income, axis = 1)
```

#### 3.8 Card_Category

```python
map8 = sns.catplot(x = 'Card_Category', kind="count", data = df, aspect = 2)
```

```python
''' 
Categorise Card_Category
0: Blue
1: Silver
2: Gold
3: Platinum
'''

def card(x):
    if x['Card_Category'] == 'Blue':
        return 0
    elif x['Card_Category'] == 'Silver':
        return 1
    elif x['Card_Category'] == 'Gold':
        return 2
    elif x['Card_Category'] == 'Platinum':
        return 3

df['Card_Category'] = df.apply(card, axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
y_df = df['Attrition_Flag']
X_df = df.drop(['CLIENTNUM', 'Attrition_Flag'], axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.2)

```

```python
X_val, X_test, y_val, y_test = train_test_split(X_test, y_test, test_size=0.5)
```

```python
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier, RandomForestClassifier, ExtraTreesClassifier, VotingClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
```

```python
mdls = [LogisticRegression(solver = 'lbfgs', max_iter = 1000),
       GaussianNB(),
       AdaBoostClassifier(),
       GradientBoostingClassifier(),
       RandomForestClassifier(),
       KNeighborsClassifier(),
       XGBClassifier(),
       ExtraTreesClassifier() 
       ]

clf = ['LogisticRegression',
      'GaussianNB',
      'AdaBoostClassifier',
      'GradientBoostingClassifier',
      'RandomForestClassifier',
      'KNeighborsClassifier',
      'XGBClassifier',
      'ExtraTreesClassifier'
      ]

i = 0
for j in mdls:
    cv = cross_val_score(j, X_train, y_train, scoring = 'accuracy', cv = KFold(shuffle = True))
    print (clf[i], '=', cv.mean())
    new_temp = [cv.mean(), clf[i]]
    i += 1
```

```python
XGBClassifier().fit(X_val, y_val)
```

```python
cv2 = cross_val_score(XGBClassifier(), X_val, y_val, scoring = 'accuracy', cv = KFold(shuffle = True))
print (cv2.mean())
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = [
  {'colsample_bytree': [0.7, 0.8, 0.9, 1],
   'learning_rate': [0.3, 0.4, 0.5, 0.6, 0.7, 0.8],
   'max_depth': [4, 5, 6, 7, 8],
   'reg_alpha':[1e-5, 1e-2, 0.1, 1, 100],
   'reg_lambda': [0.1, 1.0, 5.0, 10.0, 50.0, 100.0]}]

clf_cv = GridSearchCV(XGBClassifier(), param_grid, cv = KFold(shuffle = True), scoring='f1_macro')

clf_cv.fit(X_val, y_val)
```

```python
clf_cv.best_params_
```

```python
clf_cv.cv_results_
```

```python
cv_test = cross_val_score(XGBClassifier(colsample_bytree = 1, learning_rate= 0.3, max_depth = 5, reg_alpha = 1e-05, reg_lambda = 0.1), X_test, y_test, scoring = 'accuracy', cv = KFold(shuffle = True))
cv_test.mean()
```

```python
XGB_clf = XGBClassifier(colsample_bytree = 1, 
                          learning_rate= 0.3, 
                          max_depth = 5, 
                          reg_alpha = 1e-05, 
                          reg_lambda = 0.1).fit(X_train, y_train)
```

```python
cv_test2 = cross_val_score(XGB_clf, X_train, y_train, scoring = 'accuracy', cv = KFold(shuffle = True))
print (cv_test2.mean())
```

```python
test_pred = XGB_clf.predict(X_test)
```

```python
from sklearn import metrics
print('Precision:', metrics.precision_score(y_test, test_pred)*100,'%')
print('Accuracy:', metrics.accuracy_score(y_test, test_pred)*100,'%')
print('Recall:', metrics.recall_score(y_test, test_pred))
print('F1:', metrics.f1_score(y_test, test_pred))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, test_pred)
```
