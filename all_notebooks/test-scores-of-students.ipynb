# Prediction of Post Test of Students

```python
#import library
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, mean_absolute_error
```

## Data Extraction

```python
score = pd.read_csv('../input/predict-test-scores-of-students/test_scores.csv')
```

## Selection Data

```python
df = score[['n_student', 'posttest']]
df.head()
```

## Machine Learning Model

```python
#split data
X = df.drop('posttest', axis = 1)
y = df['posttest']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
#linear regression
linreg = LinearRegression()
linreg.fit(X_train, y_train)
print(linreg)
```

```python
#ridge regression
ridge = RidgeCV()
ridge.fit(X_train, y_train)
print(ridge)
```

```python
#lasso regression
lasso = LassoCV()
lasso.fit(X_train, y_train)
print(lasso)
```

```python
#Linear Regression prediction
linreg_pred = linreg.predict(X_test)
print(linreg_pred)
```

```python
#ridge Regression prediction
ridge_pred = ridge.predict(X_test)
print(ridge_pred)
```

```python
#lasso Regression prediction
lasso_pred = lasso.predict(X_test)
print(lasso_pred)
```

```python
#linear regression score
accuracy = linreg.score(X_test, y_test)
print('Linear Regression Score : ', accuracy)
print('Percentage : ', int(accuracy * 100), '%')
```

```python
#ridge regression score
accuracy = ridge.score(X_test, y_test)
print('Ridge Regression Score : ', accuracy)
print('Percentage : ', int(accuracy * 100), '%')
```

```python
#lasso regression score
accuracy = lasso.score(X_test, y_test)
print('Lasso Regression Score : ', accuracy)
print('Percentage : ', int(accuracy * 100), '%')
```

### Lasso Regression is the best score result

```python
#check MAE
mae = mean_absolute_error(y_test, lasso_pred)
print('Mean Absolute Error : ', mae)
```

```python
#plotting model
plt.figure(figsize=(10,6))
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, lasso.predict(X_train), color='blue')
plt.title("Average of Student ~ Post Test")
plt.xlabel("Average of Student")
plt.ylabel("Post Test")
plt.show()
```
