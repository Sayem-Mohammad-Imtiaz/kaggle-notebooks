```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

life_data = pd.read_csv("/kaggle/input/life-expectancy/lifeExpectancy_world.csv")
```

# Explore

```python
life_data.describe()
```

# Life Expectancy By Continent

```python
by_continent = pd.DataFrame()
by_continent["country"] = life_data.Entity
by_continent["life_expectancy"] = life_data.iloc[:,2:].mean(axis=1) #avg life expectancy over the years
by_continent
```

```python
continents = pd.Series(["Asia",
"Africa",
"Northern America",
"Europe",
"Antartica",
"Oceania",
"Latin America and the Caribbean"])
continents
```

```python
x = by_continent.groupby(continents).mean().reset_index().iloc[:,:1].astype(str).squeeze()
y = by_continent.groupby(continents).mean().reset_index()["life_expectancy"].astype(float)
fig, ax = plt.subplots(figsize=(25,20))
ax.bar(x,y)
ax.set_title("Average Life Expectancy Per Continent")
fig = plt.gcf()
plt.show()
fig.savefig("avg_life_per_continent.png",dpi=100)
```

# US Compared To The World

```python
x = life_data.columns[2:].astype(str).tolist()
y = life_data[life_data.Entity == "United States"].iloc[:,2:].astype(float).squeeze()

xworld = life_data.columns[2:].astype(str).tolist()
yworld = life_data[life_data.Entity == "World"].iloc[:,2:].astype(float).squeeze()


fig, ax = plt.subplots(figsize=(25,20))
ax.set_title("Average life expectancy of US over the years vs Average world life expectancy")
plt.barh(x,y, height = 0.9)
plt.barh(xworld,yworld, height = 0.9)
fig = plt.gcf()

US = mpatches.Patch(color='blue', label='United States')
W = mpatches.Patch(color='orange', label='World')
plt.legend(handles = [US,W],loc=2)

plt.show()
fig.savefig("us_vs_world_life_expectency.png",dpi=100)
```

# Northern America Compared To Europe

```python
x = life_data.columns[2:].astype(str).tolist()
y = life_data[life_data.Entity == "Northern America"].iloc[:,2:].astype(float).squeeze()

x_eu = life_data.columns[2:].astype(str).tolist()
y_eu = life_data[life_data.Entity == "Europe"].iloc[:,2:].astype(float).squeeze()

fig, ax = plt.subplots(figsize=(25,20))
ax.set_title("Average life expectancy of US over the years vs Average life expectancy in Europe")
plt.barh(x,y, height = 0.9)
plt.barh(x_eu,y_eu, height = 0.9)
fig = plt.gcf()

NA = mpatches.Patch(color='blue', label='Northern America')
EU = mpatches.Patch(color='orange', label='Europe')
plt.legend(handles = [NA,EU],loc=2)

plt.show()
fig.savefig("na_vs_eu_life_expectency.png",dpi=100)
```

# Northern America Compared To Latin America

```python
x_na = life_data.columns[2:].astype(str).tolist()
y_na = life_data[life_data.Entity == "Northern America"].iloc[:,2:].astype(float).squeeze()

x_sa = life_data.columns[2:].astype(str).tolist()
y_sa = life_data[life_data.Entity == "Latin America and the Caribbean"].iloc[:,2:].astype(float).squeeze()

fig, ax = plt.subplots(figsize=(25,20))
ax.set_title("Average life expectancy in Northern America over the years vs Average life expectancy in South America")
plt.barh(x_na,y_na, height = 0.9)
plt.barh(x_sa,y_sa, height = 0.9)
fig = plt.gcf()

NA = mpatches.Patch(color='blue', label='Northern America')
SA = mpatches.Patch(color='orange', label='Latin America')
plt.legend(handles = [NA,SA],loc=2)

plt.show()
fig.savefig("na_vs_sa_life_expectency.png",dpi=100)
```
