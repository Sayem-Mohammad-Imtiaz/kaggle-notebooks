```python
vd_data = pd.read_csv('../input/video-games-sales-data/Video_Games_Sales_as_at_22_Dec_2016.csv')
vd_data.head(10)
```

## Global sales analysis

**Year wise sales graph**

```python
df1 = vd_data.groupby(by="Year_of_Release").sum()
df1.plot.line(figsize=(10,10))
plt.ylabel("Sale in millions")
```

### Geners wise distribution of games

```python
df2 = pd.DataFrame(vd_data['Genre'].value_counts(normalize=True))
df2.plot.pie(subplots=True, figsize=(10,10), autopct='%1.1f%%')
```

### Publisher wise distribution of games


```python
x = pd.DataFrame(vd_data.groupby('Publisher')['Global_Sales'].sum())
x.sort_values(by=['Global_Sales'], inplace=True)
x = x.tail(10)
plot = x.plot.pie(y='Global_Sales', autopct='%1.1f%%', figsize=(10, 10))
plt.title("Publisher market share")
```

#### Years with most released games

```python
df3 = sns.countplot("Year_of_Release", data=vd_data)
plt.xlim([21.5,31.5])
df3.set_xticklabels(df3.get_xticklabels(), rotation=90)
plt.title("10 most frequent launch years")
plt.ylabel("Number of Games")
```

#### Years with least released games

```python
df4 = sns.countplot("Year_of_Release", data=vd_data)
plt.xlim([-0.5,9.5])
plt.ylim([0,50])
df4.set_xticklabels(df3.get_xticklabels(), rotation=90)
plt.title("10 least frequent launch years")
plt.ylabel("Number of Games")
```

### Top 10 Publishers(global sales)

```python
df5 = pd.DataFrame(vd_data.groupby('Publisher')['Global_Sales'].sum())
df5.sort_values(by=['Global_Sales'], inplace=True)
df5 = df5.tail(10)
df5.plot.pie(subplots=True, autopct='%1.1f%%', figsize = (10,10))
plt.title("Top 10 Publishers Globally")
```

### Top 10 Publishers area wise

```python
df6 = pd.DataFrame(vd_data.groupby('Name')['NA_Sales'].sum())
df6.sort_values(by=['NA_Sales'], inplace=True)
df6 = df6.tail(10)
df6.plot.pie(subplots=True, autopct='%1.1f%%', figsize = (10,10))
plt.title("Top 10 Publishers in North America")
```

```python
df7 = pd.DataFrame(vd_data.groupby('Name')['EU_Sales'].sum())
df7.sort_values(by=['EU_Sales'], inplace=True)
df7 = df7.tail(10)
df7.plot.pie(subplots=True, autopct='%1.1f%%', figsize = (10,10))
plt.title("Top 10 Publishers in Europe")
```

```python
df8 = pd.DataFrame(vd_data.groupby('Name')['JP_Sales'].sum())
df8.sort_values(by=['JP_Sales'], inplace=True)
df8 = df8.tail(10)
df8.plot.pie(subplots=True, autopct='%1.1f%%', figsize = (10,10))
plt.title("Top 10 Publishers in Japan")
```
