```python
for i in range(0,len(df)):
    print(df.iloc[i,8])
    if ','in str(df.iloc[i,8]):
        cl=df.iloc[i,8].split(',')
        df.iloc[i,8]=int(cl[0])
        df.iloc[i,11]=int(cl[1])
        if len(cl)>2:
            df.iloc[i,12]=int(cl[2])
        elif df.iloc[i,8]>100:
            df.iloc[i,11]=df.iloc[i,8]%100
            df.iloc[i,8]=df.iloc[i,8]//100
```

```python
df.groupby(by=0).count()
city=['Ghaziabad','Kanpur','Lucknow']
total_crimes=np.array([167,97,202])
fig=plt.figure(figsize=(10,5))
plt.bar(city,total_crimes,color='violet',width=0.4)
plt.xlabel("CITIES")
plt.ylabel("Total Crimes Commited")
plt.title("CRIMES COMMITTED IN CITIES")
plt.show()
```

 Number of crimes commited is maximum in Lucknow.

```python
city=['Ghaziabad','Kanpur','Lucknow']
crimes=np.array([167,97,202])
plt.pie(crimes,labels=city,radius=2)
plt.show()
```

```python
sums = [df[5].sum(),df[6].sum(),df[7].sum()]
victims=['Child','Male','Female']
plt.pie(sums,labels=victims,shadow=True,startangle=90)
plt.legend()
plt.legend(title="KIDNNAPING")
plt.show()
```

 Observation- From above pie chart we can see that the number of child victims are higher than male and female in kidnapping case.
