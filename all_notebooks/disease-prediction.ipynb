```python
df_train= pd.read_csv('../input/disease-prediction-using-machine-learning/Training.csv')
df_train=df_train.drop('Unnamed: 133',axis=1)
df_train.head(10)
```

```python
X_train= df_train.drop('prognosis',axis=1)
y_train= df_train['prognosis']
print(X_train.shape)
print(len(y_train.unique()))
```

```python
y_train=pd.get_dummies(y_train)
y_train.head()
```

#Create a NN model

```python
model = Sequential()
model.add(Dense(20,input_dim=132,activation='relu',name='layer_1'))
model.add(Dense(40,activation='relu',name='layer_2'))
model.add(Dense(50,activation='relu',name='layer_3'))
model.add(Dense(41,activation='softmax',name='output_layer'))
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(X_train,y_train,epochs=10,verbose=1,shuffle=True)

```

<h1> Test Data Study </h1>

```python
df_test= pd.read_csv('../input/disease-prediction-using-machine-learning/Testing.csv')
df_test.head()
X_test= df_test.drop('prognosis',axis=1)
y_test= df_test['prognosis']
y_test=pd.get_dummies(y_test)

import numpy as np

y_pred = model.predict(X_test)

print(model.evaluate(X_test,y_test))



```
