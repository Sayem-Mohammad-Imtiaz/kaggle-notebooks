```python
#Find the different types of breast cancer genes expressed
breast_cancer_types = np.unique(list(chain(*breast_cancer_data['type'].map(lambda x: x.split('|')).tolist())))
breast_cancer_types = [x for x in breast_cancer_types if len(x)>0]
print('Types ({}): {}'.format(len(breast_cancer_types), breast_cancer_types))
```

The genes expreesed in the breast tissues are grouped into HER, basal, cell-line, luminal_A, luminal_B and normal. "normal" is said to be normal breast gene expression. I.e non-cancerous

```python
#plotting the value counts of the different types of breast cancers, with "Normal" refering to normal breast tissue
ax= sns.countplot(breast_cancer_data['type'])
ax.set(ylabel = 'Count')
ax.set(xlabel = 'Type of breast cancer')
ax.set(title = 'Statistics of the different breast cancer types')

```
