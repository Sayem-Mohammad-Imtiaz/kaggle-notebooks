```python
us_accidents = pd.read_csv('/kaggle/input/us-accidents/US_Accidents_Dec20_Updated.csv')
```

First to take a glance at the column start time of the dataset.

```python
start_time = us_accidents['Start_Time']
start_time.describe()
```

This column is type of object and it would be better to process the data when it is converted to datetime. Besides, it makes much more sense to get the hour of the datetime.

```python
start_time = pd.to_datetime(start_time)
```

```python
start_time = start_time.dt.hour
```

```python
start_time.head()
```

```python
acc_amount = start_time.value_counts()
```

```python
print('Most car accidents happens at %s.' %acc_amount.index[0])
```

It is time to visualise amount of car accidents happending at which time.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style('whitegrid')
sns.set_context('paper')

us_accidents["accident_time_hour"] = start_time

sns.countplot(x="accident_time_hour", data=us_accidents)
```
