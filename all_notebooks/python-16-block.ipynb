```python
df = pd.read_csv('/kaggle/input/russian-demography/russian_demography.csv')
df
```

```python
import statistics as stat

median_birth_rate=np.median(df['birth_rate'])
print("Медиана рождаемости: ", median_birth_rate)

median_death_rate=np.median(df['death_rate'])
print("Медиана смертности:", median_death_rate)

median_gdw=np.median(df['gdw'])
print("Медиана BBП: ", median_gdw)

mode_birth_rate = stat.mode(df['birth_rate'])
print ("Мода рождаемости: ", mode_birth_rate)

mode_death_rate=stat.mode(df['death_rate'])
print("Мода смертности: ", mode_death_rate)

#mode_gdw=stat.mode(df['gdw'])  (Статистическая ошибка: нет уникального режима; найдено 3 одинаково распространенных значения)
#print("Мода ВВП: ", mode_gdw)

mean_birth_rate = np.mean(df['birth_rate'])
print("Средняя рождаемость: ", mean_birth_rate)

mean_death_rate=np.mean(df['death_rate'])
print("Средняя смертность: ", mean_death_rate)

mean_gdw=np.mean(df['gdw'])
print("Средняя ВВП: ", mean_gdw)

max_birth_rate=np.max(df['birth_rate'])
print("Максимум рождаемости: ", max_birth_rate)

max_death_rate=np.max(df['death_rate'])
print("Максимум смертности: ", max_death_rate)

max_gdw=np.max(df['gdw'])
print("Максимум ВВП: ", max_gdw)

min_birth_rate=np.min(df['birth_rate'])
print("Минимум рождаемости: ", min_birth_rate)

min_death_rate=np.min(df['death_rate'])
print("Минимум смертности: ", min_death_rate)

min_gdw=np.min(df['gdw'])
print("Минимум ВВП: ", min_gdw)
```

```python
df['new']=df.max(axis = 1) - df.min(axis = 1)
df
```
