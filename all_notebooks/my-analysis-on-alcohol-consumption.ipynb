```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('../input/alcohol-consumption/gapminder_alcohol.csv')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['alcconsumption'].fillna(df['alcconsumption'].mean(), inplace=True)
df['incomeperperson'].fillna(df['incomeperperson'].mean(), inplace=True)
df['suicideper100th'].fillna(df['suicideper100th'].mean(), inplace=True)
df['employrate'].fillna(df['employrate'].mean(), inplace=True)
df['urbanrate'].fillna(df['urbanrate'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
plt.scatter(df.alcconsumption, df.suicideper100th)
plt.xlabel('alcohol consumption')
plt.ylabel('suicide per 100th')
plt.title('alcohol consuption - suicide',size=20)
plt.show()
```

```python
plt.scatter(df.employrate, df.incomeperperson)
plt.xlabel('employe rate')
plt.ylabel('income per person')
plt.title('employe rate-income per person',size=20)
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.scatter(df.country, df.employrate,label='employrate')
plt.scatter(df.country,df.suicideper100th,label='sucide')
plt.xlabel('country')
plt.ylabel('employrate / sucide')
plt.title('country-employe rate',size=20)
plt.legend()
plt.show()
```

```python
df.hist()
plt.show()
```

```python
df["suicideper100th"].plot(kind = 'hist',bins = 200,figsize = (6,6))
plt.title("Sucide per 100")
plt.xlabel("Sucide")
plt.ylabel("Frequency")
plt.show()
```

```python
y=np.array([df['incomeperperson'].min(),df['incomeperperson'].mean(),df['incomeperperson'].max()])
x=['low','average','high']
plt.bar(x,y,label='sdfas',color='r',width=0.3)
plt.xlabel('level')
plt.ylabel('income per person')
plt.show()
```

```python
y=np.array([df['employrate'].min(),df['employrate'].mean(),df['employrate'].max()])
y1=np.array([df['alcconsumption'].min(),df['alcconsumption'].mean(),df['alcconsumption'].max()])
x=['low','average','high']
plt.bar(x,y,label='employrate')
plt.bar(x,y1,label='alcconsumption',bottom=y)
plt.xlabel('level')
plt.ylabel('employrate / alcconsumption')
plt.legend()
plt.show()
```

```python
alc=['alcconsumption', 'incomeperperson', 'suicideper100th', 'employrate', 'urbanrate']

a=5
b=3
c=1

fig = plt.figure(figsize=(20,28))

for i in alc:
    plt.subplot(a, b, c)
    plt.title(f'{i} (dist), subplot: {a,b,c}')
    plt.xlabel(i)
    sns.distplot(df[i])
    c = c + 1

    plt.subplot(a, b, c)
    plt.title(f'{i} (box), subplot: {a,b,c}')
    plt.xlabel(i)
    sns.boxplot(x = df[i],color='r')
    c = c + 1

    plt.subplot(a, b, c)
    plt.title(f'{i} (scatter), subplot: {a,b,c}')
    plt.xlabel(i)
    sns.scatterplot(data = df, x = i, y = 'urbanrate',color='g')
    c = c + 1

plt.show()
```
