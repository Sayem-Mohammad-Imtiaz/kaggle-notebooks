```python
import pandas as pd
from pandas.io.json import json_normalize
import warnings
warnings.filterwarnings("ignore")
from sklearn.feature_extraction.text import CountVectorizer
import random
from wordcloud import WordCloud 
import matplotlib.pyplot as plt
import seaborn as sns
import json_lines
from tqdm.notebook import tqdm
```

```python
data=pd.read_csv('../input/covid19-tweets/covid19_tweets.csv')
data
```

```python
import re
import nltk
nltk.download("stopwords")
nltk.download('punkt')
from nltk import word_tokenize,sent_tokenize
nltk.download('wordnet')
import nltk as nlp
```

```python
df=data['text']
cult_list=[]
for cult in tqdm(df):
    cult=re.sub("[^a-zA-z]"," ",str(cult))
    cult=cult.lower()
    cult=nltk.word_tokenize(cult)
    lemma=nlp.WordNetLemmatizer()
    cult=[lemma.lemmatize(word) for word in cult]
    cult=" ".join(cult)
    cult_list.append(cult)
```

```python
max_features=400
count_vectorizer=CountVectorizer(max_features=max_features,stop_words="english")
sparce_matrix=count_vectorizer.fit_transform(cult_list).toarray()
data3=pd.DataFrame(count_vectorizer.get_feature_names(),columns=["Words"])
```

```python
fig,axs = plt.subplots(3,3,figsize=(16,12))
for i in range(9):
    r=i//3
    c=i%3
    random.seed(2020+i)
    random.shuffle(data3.Words)
    wordcloud=WordCloud(background_color="black",width=1024,height=768).generate(" ".join(data3.Words))    
    ax=axs[r][c].axis("off")   
    ax=axs[r][c].imshow(wordcloud)
plt.show()
```
