```python
# подгрузим библиотеки
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# прочтем файл и просмотрим его
df = pd.read_csv('../input/russia-alcohol/russia_alcohol.csv')
df
```

```python
# медиана

median_wine = df.wine.median()
print("Медиана по покупке вина: ", median_wine)

median_beer = df.beer.median()
print("Медиана по покупке пива: ", median_beer)

median_vodka = df.vodka.median()
print("Медиана по покупке водки: ", median_vodka)

median_champagne = df.champagne.median()
print("Медиана по покупке шампанского: ", median_champagne)

median_brandy = df.brandy.median()
print("Медиана по покупке бренди: ", median_brandy)
```

```python
# мода

mode_wine = df.wine.mode()
print("Мода по покупке вина: ", mode_wine)

mode_beer = df.beer.mode()
print("Мода по покупке пива: ", mode_beer)

mode_vodka = df.vodka.mode()
print("Мода по покупке водки: ", mode_vodka)

mode_champagne = df.champagne.mode()
print("Мода по покупке шампанского: ", mode_champagne)

mode_brandy = df.brandy.mode()
print("Мода по покупке бренди: ", mode_brandy)
```

```python
# средняя

mean_wine = df.wine.mean()
print("Среднее по покупке вина: ", mean_wine)

mean_beer = df.beer.mean()
print("Среднее по покупке пива: ", mean_beer)

mean_vodka = df.vodka.mean()
print("Среднее по покупке водки: ", mean_vodka)

mean_champagne = df.champagne.mean()
print("Среднее по покупке шампанского: ", mean_champagne)

mean_brandy = df.brandy.mean()
print("Среднее по покупке бренди: ", mean_brandy)
```

```python
# минимум

min_wine = np.min(df['wine'])
print("Минимум по покупке вина: ", min_wine, '\nРегион:', df.region[df.wine == min_wine])

min_beer = np.min(df['beer'])
print("\nМинимум по покупке пива: ", min_beer, '\nРегион:', df.region[df.beer == min_beer])

min_vodka = np.min(df['vodka'])
print("\nМинимум по покупке водки: ", min_vodka, '\nРегион:', df.region[df.vodka == min_vodka])

min_champagne = np.min(df['champagne'])
print("\nМинимум по покупке шампанского: ", min_champagne, '\nРегион:', df.region[df.champagne == min_champagne])

min_brandy=np.min(df['brandy'])
print("\nМинимум по покупке бренди: ", min_brandy, '\nРегион:', df.region[df.brandy == min_brandy])
```

```python
# максимум

max_wine = np.max(df['wine'])
print("Максимум по покупке вина: ", max_wine, '\nРегион:', df.region[df.wine == max_wine])

max_beer = np.max(df['beer'])
print("\nМаксимум по покупке пива: ", max_beer, '\nРегион:', df.region[df.beer == max_beer])

max_vodka = np.max(df['vodka'])
print("\nМаксимум по покупке водки: ", max_vodka, '\nРегион:', df.region[df.vodka == max_vodka])

max_champagne = np.max(df['champagne'])
print("\nМаксимум по покупке шампанского: ", max_champagne, '\nРегион:', df.region[df.champagne == max_champagne])

max_brandy = np.max(df['brandy'])
print("\nМаксимум по покупке бренди: ", max_brandy, '\nРегион:', df.region[df.brandy == max_brandy])
```

```python
# Создать новый столбец, в котором разместить разницу между максимумом и минимумом

df['max-min'] = df.max(axis = 1) - df.min(axis = 1)
df
```

```python
'''Разбить данные на несколько блоков (не менее 3, приблизительно равных по размеру) по 
некоторому критерию (критерий определить самостоятельно).'''
# разделим на президентов

elcyn = df[(df['year'] <= 1999)&(df['year'] >= 1996)]
putin = df[(df['year'] >= 2000)&(df['year'] <= 2007)]
medvedev = df[(df['year'] >= 2008)&(df['year'] <= 2011)]
putin_again = df[(df['year'] >= 2012)&(df['year'] <= 2016)]

elcyn
```
