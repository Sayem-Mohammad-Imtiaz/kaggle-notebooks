```python
import pandas as pd
covid = pd.read_csv("../input/covid19-coronavirus/2019_nCoV_data.csv")
```

```python
covid["Country"]
```

```python
covid["Deaths"]
```

```python
countriesList = []
for country in covid["Country"]:
    for unique in country.split(","):
        if unique not in countriesList:
            countriesList.append(unique)
print(countriesList)
```

```python
countryTotal = {}    
countryCount = {}    

for country in countriesList:
    countryTotal[country] = 0
    countryCount[country] = 0
print(countryTotal)
print(countryCount)
```

```python
for idx, countries in enumerate(covid["Country"]):
    
    for country in countries.split(","):
        countryTotal[country] += covid["Deaths"][idx] 
        countryCount[country] += 1
print(countryTotal)
print("******")
print(countryCount)
```

```python
countryAverage = {}

for country in countryTotal.keys():
    countryAverage[country]= countryTotal[country] / countryCount[country]
print(countryAverage)
```

```python
resultCountry = "temp"
resultDeaths = 0
for country in countryAverage.keys():
    if countryAverage[country] > resultDeaths:
        resultCountry = country
        resultDeaths = countryAverage[country]
        
print( resultCountry +  " HAS THE MOST DEATH RATE WHICH IS "  +   str(round(resultDeaths,2)) +  " BECAUSE OF COVID19 CORONAVIRUS")
```
