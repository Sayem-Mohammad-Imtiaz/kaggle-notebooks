```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
import nltk 
import string
import re
%matplotlib inline
pd.set_option('display.max_colwidth', 100)
```

```python
# Load dataset
def load_data():
    data = pd.read_csv('../input/covid19-news-in-bali/m_berita.csv')
    return data
```

```python
berita_df = load_data()
berita_df.head()
```

```python
print('Dataset size:',berita_df.shape)
print('Columns are:',berita_df.columns)

```

```python
berita_df.info()
```

```python
df  = pd.DataFrame(berita_df[['id', 'berita','judul']])

from wordcloud import WordCloud, STOPWORDS , ImageColorGenerator
# Start with one review:
judul_All = " ".join(str(review) for review in df.judul)
berita_All = " ".join(str(review) for review in df.berita)

fig, ax = plt.subplots(2, 1, figsize  = (30,30))
# Create and generate a word cloud image:
wordcloud_ALL = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(berita_All)
wordcloud_JUDUL = WordCloud(max_font_size=50, max_words=100, background_color="black").generate(judul_All)
# Display the generated image:
ax[0].imshow(wordcloud_ALL, interpolation='bilinear')
ax[0].set_title('Isi Berita', fontsize=30)
ax[0].axis('off')
ax[1].imshow(wordcloud_JUDUL, interpolation='bilinear')
ax[1].set_title('Judul', fontsize=30)
ax[1].axis('off')
```
