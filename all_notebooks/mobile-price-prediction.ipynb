# **Data PreProcessing**

```python
dataset = pd.read_csv('/kaggle/input/mobile-price-classification/train.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().sum()
```

```python
dataset = dataset.drop_duplicates()
dataset.info()
```

```python
Y = dataset['price_range']
Y
```

```python
X = dataset.drop('price_range',1)
X
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from sklearn.metrics import precision_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

from sklearn.pipeline import Pipeline

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

from sklearn.feature_selection import f_classif, chi2
from sklearn.feature_selection import SelectKBest, SelectPercentile
```

**Split Data**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size = 0.33, random_state=42)
```

**Feature Selection**

```python
corr=dataset.corr()
fig = plt.figure(figsize=(15,12))
r = sns.heatmap(corr, cmap='Purples')
r.set_title('Correlations')
```

```python
corr.sort_values(by=['price_range'],ascending=False).iloc[0].sort_values(ascending=False)
```

```python
sel_ = SelectKBest(f_classif, k=5).fit(X_train.fillna(0), Y_train)
X_train.columns[sel_.get_support()]
```

```python
X_train = sel_.transform(X_train.fillna(0))
X_train.shape
```

```python
X_test = sel_.transform(X_test.fillna(0))
X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
Y_pred=dt.predict(X_test)
print(classification_report(Y_test,Y_pred))
accDT=accuracy_score(Y_test,Y_pred)
print(accDT)
```

```python
knnm = KNeighborsClassifier(n_neighbors=10)
knnm.fit(X_train,Y_train)
knnm.score(X_test,Y_test)
y_knn_pred = knnm.predict(X_test)
print(classification_report(Y_test,y_knn_pred))
accKNN = accuracy_score(Y_test,y_knn_pred)
print(accKNN)
```

```python
gnb = GaussianNB()
y_pred = gnb.fit(X_train, Y_train).predict(X_test)
print(classification_report(Y_test,y_pred))
accNB = accuracy_score(Y_test,y_knn_pred)
print(accNB)
```

```python
result = [accDT, accKNN, accNB]
classtrs= ['DecisionTree', 'KNN', 'NaiveBayes']

plt.bar(classtrs, result)
plt.xlabel('Classifier')
plt.ylabel('Accuracy Score')
plt.title('Non-standardized Accuracy Scores')
plt.show()
```
