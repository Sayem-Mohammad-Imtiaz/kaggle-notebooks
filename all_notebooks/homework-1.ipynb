```python
x = data_2[data_2['Country name'] == 'Turkey']['year']
y = data_2[data_2['Country name'] == 'Turkey']['Freedom to make life choices']
z = data_2[data_2['Country name'] == 'Greece']['year']
q = data_2[data_2['Country name'] == 'Greece']['Freedom to make life choices']
k = data_2[data_2['Country name'] == 'Georgia']['year']
l = data_2[data_2['Country name'] == 'Georgia']['Freedom to make life choices']
fig = go.Figure()
fig.add_trace(go.Scatter(x=x, y=y, name='Freedom to make life choices',
                         line=dict(color='green', width=4)))
fig.add_trace(go.Scatter(x= z, y=q, name ='Freedom to make life choices',
                         line=dict(color='royalblue', width=4)))
fig.add_trace(go.Scatter(x=k, y=l, name='High 2007',
                         line=dict(color='firebrick', width=4)))
```

Line Plots of freedom to make life choices of Turkey, Greece and Georgia.


```python
data_2.plot(kind= 'scatter', x = 'year', y = 'Freedom to make life choices', alpha = 0.5, color = 'green')
plt.ylabel('Freedom to make life choices')
plt.title('Year')
```

According to the graph we can see Freedom to make life choices of world is shows a decreasing trend from 2005 to 2012 and than increasing trend from 2012 to 2020. Although the reason for the decline from 2005 to 2012 depends on many factors, the main factor can be seen as the environment of economic crisis around the world.
