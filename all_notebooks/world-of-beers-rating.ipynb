```python
df = pd.read_csv("/kaggle/input/world-of-beers/beer_list.csv",
                  engine="python", sep=",")
df.head()
```

```python
df.isnull().sum()
```

```python
#Code by Rashik Rahman  https://www.kaggle.com/rashikrahmanpritom/insights-on-disney-movies-gross

# Sort data by Styles in descending order 
Style_desc = df.sort_values(by='Style', ascending=False) 

# Display the top 10 Styles 
Style_desc.head(10)
```

```python
#Code by Rashik Rahman https://www.kaggle.com/rashikrahmanpritom/insights-on-disney-movies-gross/notebook

# Extract year from release_date and store it in a new column
df['release_year'] = pd.DatetimeIndex(df['Date']).year

# Compute mean of adjusted gross per genre and per year
group = df.groupby(['Style','release_year']).mean()

# Convert the GroupBy object to a DataFrame
Style_yearly = group.reset_index()

# Inspect genre_yearly 
print(Style_yearly.head())


# Plot the data  

g=sns.relplot(kind='line', x='release_year', y='Catalog', hue='Style', data=Style_yearly)
g.fig.set_size_inches(10,10)
plt.title('Beer Style trend')
```

```python
df = df.rename(columns={'Brewery/Beer':'brewery'})
```

```python
#Code by Siti K https://www.kaggle.com/khotijahs1/2020-indonesia-university-rank/comments

#The Best 20 Beer's Style by Score
top_beer = df.sort_values(by='Score', ascending=False)[:20]
figure = plt.figure(figsize=(10,6))
sns.barplot(y=top_beer.brewery, x=top_beer.Score)
plt.xticks()
plt.xlabel('Score')
plt.ylabel('Brewery')
plt.title('Top 20 Breweries')
plt.show()
```

```python
#word cloud
from wordcloud import WordCloud, ImageColorGenerator
text = " ".join(str(each) for each in df.brewery)
# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=200,colormap='Pastel1', background_color="black").generate(text)
plt.figure(figsize=(10,6))
plt.figure(figsize=(15,10))
# Display the generated image:
plt.imshow(wordcloud, interpolation='Bilinear')
plt.axis("off")
plt.figure(1,figsize=(12, 12))
plt.show()
```
