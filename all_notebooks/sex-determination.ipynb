```python
test=pd.read_csv("../input/male-female-height-and-weight/Test set.csv")
train=pd.read_csv('../input/male-female-height-and-weight/Training set.csv')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(test.Height,test.Weight, hue=test.Sex)
```

```python
test['Sex'].replace('Female',0, inplace=True)
test['Sex'].replace('Male',1, inplace=True)
train['Sex'].replace('Female',0, inplace=True)
train['Sex'].replace('Male',1, inplace=True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn15=KNeighborsClassifier(n_neighbors=21)
```

```python
knn15.fit(train.drop(['Sex'],axis=1),train['Sex'])
```

```python
predicted=knn15.predict(test.drop(['Sex'],axis=1))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(test['Sex'],predicted))
```

```python
accu=accuracy_score(test['Sex'],predicted)
```

```python
error_rate = []
X_train=train.drop(['Sex'],axis=1)
X_test=test.drop(['Sex'],axis=1)
y_train=train['Sex']
y_test=test['Sex']

for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))

plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_test, predicted)
```

```python
roc_auc_score(y_test, y_predicted)
```

```python
from sklearn import svm
clf = svm.SVC(kernel='linear') 
clf.fit(X_train, y_train)
y_predicted = clf.predict(X_test)
```
