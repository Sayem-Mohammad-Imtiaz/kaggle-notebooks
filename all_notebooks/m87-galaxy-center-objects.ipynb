```python
#Reading data
import pandas as pd
import numpy as np
import glob

interesting_files = glob.glob("/kaggle/input/m87-galaxy-center-objects-csv-file-esa/M87_CENTER_OBJECTS/*.csv") 
df = pd.concat((pd.read_csv(f, header = 0) for f in interesting_files))
df.to_csv("output.csv")
```

```python
df.head()
```

```python
#dropping object type columns
df = df.select_dtypes(exclude=['object'])
# print(df)
```

```python
df.info()
```

```python
#list of all column name
list(df.columns.values)
```

```python
# Create correlation matrix
corr_matrix = df.corr().abs()

# Select upper triangle of correlation matrix
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))

# Find features with correlation greater than 0.95
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]

# Drop features 
df.drop(to_drop, axis=1, inplace=True)
```

```python
max_nan_value_count = 130
df = df.drop(df.columns[df.apply(lambda 
col: col.isnull().sum() > max_nan_value_count)], axis=1)
```
