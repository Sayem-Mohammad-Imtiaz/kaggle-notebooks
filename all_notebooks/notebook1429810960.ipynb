```python
import pandas as pd
import numba as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("../input/kerala-assembly-election-history/Election dataset1.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
plt.figure(figsize=(10,7))
plt.style.use("fivethirtyeight")
sns.lineplot(x='Year', y='Total Electors',data=data , label="Total Electors")
sns.lineplot(x='Year', y='Total Votes',data=data ,label="Total Votes")
```

```python
plt.figure(figsize=(30,7))
sns.countplot(data['Party'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.countplot(data['Type'])
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.style.use("ggplot")
sns.lineplot(x='Year', y='Margin',data=data , label="Margins")
```

```python
plt.figure(figsize=(10,7))
plt.style.use("ggplot")
sns.lineplot(x='Year', y='Margin %',data=data , label="Margin %")
```

```python
plt.figure(figsize=(20,7))
data.groupby(['Party'])['Total Votes'].sum().sort_values().plot()
```

```python
plt.figure(figsize=(20,7))
data.groupby(['Party'])['Total Electors'].sum().sort_values().plot()
```

```python
data.groupby(['Type'])['Total Votes'].sum().sort_values().plot()
```
