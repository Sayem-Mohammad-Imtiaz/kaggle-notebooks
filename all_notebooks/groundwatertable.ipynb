```python
df=pd.read_csv('https://data.wra.gov.tw/Service/OpenData.aspx?format=csv&id=C79B9CF3-122E-43B6-A436-D7F6B40C0D3A')
df
```

```python
#改column名稱
df.columns=['年平均水位','年最高日水位','瞬間最高水位','年最低日水位','瞬間最低水位','年最高日水位發生日','年最低日水位發生日','瞬間最高水位發生日時','瞬間最低水位發生日時','井號','西元年']
df
```

```python
#將年水位差(年最高日水位-年最低日水位)加入欄位
df['年水位差']=df.年最高日水位 - df.年最低日水位
df
```

```python
#將年水位差從高到低排列
df_sort=df.sort_values(by='年水位差',ascending=False)
df_sort.loc[:,['井號','西元年','年水位差']]
```

```python
#年平均水位至少3.5的資料內容
year_ave = df.loc[df.年平均水位 >= 3.5]
year_ave
```

```python
#依'西元年'做分組並計算出現多少次
year_count=df.groupby('西元年').西元年.count()
year_count
```

```python
#依'西元年'分組後，取得2000年的內容
y2000 = df.groupby('西元年').get_group(2000)
y2000
```

```python
#將'井號'進行分組，計算出現次數並排序
number_count = df.groupby('井號').井號.count().sort_values()
number_count
```
