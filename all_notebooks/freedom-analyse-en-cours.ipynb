```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("../input/index.csv")
df.head()
```

```python
print("Score Max en 2015 : ", df['2015 Score'].max())
print("Score Max en 2012 : ", df['2014 Score'].max())
print("Score Max en 2011 : ", df['2014 Score'].max())
print("Score Max en 2001 : ", df['2014 Score'].max())
#sum(pd.isnull(df['2015 Status']))
```

```python
Worst = df[['Country','2015 Score']].sort_values(by='2015 Score', ascending=False).head(15)
Worst.plot(x="Country", kind="bar")
plt.title("Les 15 derniers pays sur l'echelle du score de liberté de la presse ")
```

```python
Best = df[['Country','2015 Score']].sort_values(by='2015 Score', ascending=True).head(15)
Best.plot(x="Country", kind="bar")
plt.title("Les 15 premiers pays sur l'echelle du score de liberté de la presse ")
```

```python
s_2001 = df.groupby('2001 Status')['Country'].count().rename("2001")
s_2002 = df.groupby('2002 Status')['Country'].count().rename("2002")
s_2003 = df.groupby('2003 Status')['Country'].count().rename("2003")
s_2004 = df.groupby('2004 Status')['Country'].count().rename("2004")
s_2005 = df.groupby('2005 Status')['Country'].count().rename("2005")
s_2006 = df.groupby('2006 Status')['Country'].count().rename("2006")
s_2007 = df.groupby('2007 Status')['Country'].count().rename("2007")
s_2008 = df.groupby('2008 Status')['Country'].count().rename("2008")
s_2009 = df.groupby('2009 Status')['Country'].count().rename("2009")
s_2010 = df.groupby('2010 Status')['Country'].count().rename("2010")
s_2011 = df.groupby('2011 Status')['Country'].count().rename("2011")
s_2012 = df.groupby('2012 Status')['Country'].count().rename("2012")
s_2013 = df.groupby('2013 Status')['Country'].count().rename("2013")
s_2014 = df.groupby('2014 Status')['Country'].count().rename("2014")
s_2015 = df.groupby('2015 Status')['Country'].count().rename("2015")

s = pd.concat([s_2001,s_2002,s_2003,s_2004,s_2005,s_2006,s_2007,s_2008,s_2009,s_2010,s_2011,s_2012,
              s_2013,s_2014,s_2015], join='inner', axis=1)

s.transpose().plot(kind='line',figsize=(6, 3))
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.title("Evolution de la liberté de la presse entre 2001 et 2015")
```

**Les plus hauts taux de variation entre 2010 et 2015**

```python
year2010=df.groupby('Country')['2010 Score'].sum()
year2015=df.groupby('Country')['2015 Score'].sum()

Evolution1 = pd.DataFrame(((year2015-year2010)/year2010)*100)
Evolution1.columns = ['Evolution_2010_2015']
Evol1_NEG = Evolution1.sort_values(by='Evolution_2010_2015').head(15)
Evol1_POS = Evolution1.sort_values(by='Evolution_2010_2015', ascending=False).head(15)

Evol1 = pd.concat([Evol1_NEG,Evol1_POS])  
Evol1b=Evol1.sort_values(by='Evolution_2010_2015', ascending=False)
Evol1b.plot( kind="bar")
plt.title("Les 30 pays où le score de non liberté de la presse a fortement évolué [à la hausse ou à la baisse] depuis l'année 2010")


#Evolution1.index = Evolution1.index.values
#df['Evolution1'] = Evolution1
```

**A DataFrame per year**

```python
years = range(2001, 2016)
strings = ['Legal Score', 'Political Score', 'Economic Score', 'Score', 'Status']

dfs = []
for y in years:
    cols = ['Country'] + list(map((lambda s: '%d %s' % (y, s)), strings))
    d = pd.DataFrame(columns=cols, data=df[cols])
    d[str(y)] = y
    d.rename(columns={str(y): 'Year'}, inplace=True)
    dfs.append(d)

dfs[14].head()
```

**Final Join**

```python
df_rows = pd.DataFrame()
for d in dfs:
    d.columns = ['Country'] + strings + ['Year']
    df_rows = df_rows.append(d)

print(df_rows.shape)
df_rows.tail()
```
