# > *****Taking mean values of 'First Tooltip' of each year*

```python
def plot_charts(file_name):
      
        df=pd.read_csv('/kaggle/input/who-worldhealth-statistics-2020-complete/'+file_name+'.csv')

        fig = px.choropleth(df, locations="Location",locationmode='country names',animation_frame='Period',
                        color="First Tooltip",title=file_name,
                        hover_name="Location",hover_data=['Indicator'] 

                       )
        fig.show()



        fig=px.bar(x=df.groupby('Location')['First Tooltip'].mean().sort_values()[::3].index,
                       y=df.groupby('Location')['First Tooltip'].mean().sort_values()[::3].values,
                       title=file_name,
                  labels={'x':'Country','y':'First Tooltip'}
            )

        fig.update_layout(xaxis_tickangle=-90)

        fig.show()
```
