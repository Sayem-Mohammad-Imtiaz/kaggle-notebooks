**Decision Tree Regression**(To predict salaries based on previous job position)

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing dataset
ds =pd.read_csv('../input/positionsalaries-data/position-salaries data.csv')
ds.head()
```

Position and Level column, both show the position of the individual's previous job, so to avoid categorical data in position column, we use only Level column as the independent variable.

```python
x = ds.iloc[:, 1:-1].values
y = ds.iloc[:, -1].values
print(x)
print(y)
```

We do not split the data since in the dataset there are only 10 values. So, splitting the data further into test and train would decrease the accuracy.


```python
#training the model
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
dt.fit(x,y)
```

```python
#visualising predicted results
x_new = np.arange(min(x), max(x), 0.1).reshape(-1, 1)
plt.scatter(x, y, c = 'red')
plt.plot(x_new, dt.predict(x_new))
plt.title('Decision Tree Regression')
plt.xlabel('Position Level')
plt.ylabel('Salary')
```

```python
#checking accuracy
from sklearn.metrics import r2_score
r2_score(y, dt.predict(x))
```

**Accuracy = 100%**
