```python
#Importing Necessary Modules
import numpy as np 
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
from matplotlib import colors as cls
plt.style.use('fivethirtyeight')

%matplotlib inline
```

# **Load the Data**

```python
df = pd.read_csv('../input/covid19-confirmed-cases-kerala/Covid 19 Confirmed Cases-Kerala.csv')
df.shape
```

```python
df.isna().sum()
```

```python
df.tail()
```

# **Converting Date column to pandas DateTime**

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
plt.figure(figsize=(15,10)) 

sns.lineplot(data=df,x='Date',y='Confirmed')
plt.title('Kerala Covid-19 Confirmed Cases')

plt.xticks(rotation=45)
plt.show()
```

```python
df[['Confirmed']].describe()
```

# **Simple Moving Average**

```python
df['Moving Average'] = df['Confirmed'].rolling(window=4).mean()
df.head()
```

```python
df.tail()
```

```python
def second_plot(y2,color2,title=None):
    plt.figure(figsize=(10,5))
    
    plt.plot(df['Date'], df['Confirmed'], color='blue', label='Confirmed')
    plt.plot(df['Date'], df[y2], color=color2, label=y2)
    
    plt.title(title)
    plt.xlabel('Dates')
    plt.ylabel('Confirmed Cases')
    plt.legend(loc='upper left')
    plt.show()
```

```python
second_plot('Moving Average','orange','Confirmed vs Simple Moving Average')
```

# **Weighted Moving Average**

```python
weights = np.arange(1,5)
df['Weighted Moving Average'] = df['Confirmed'].rolling(4).apply(lambda confirmed: np.dot(confirmed, weights)/weights.sum(), raw=True)
```

```python
second_plot('Weighted Moving Average','Yellow','Confirmed vs Weighted Moving Average')
```

# **Exponential Moving Average**

```python
df['Exponential Moving Average'] = df['Confirmed'].ewm(span=4, adjust=False).mean()
```

```python
second_plot('Exponential Moving Average','red','Confirmed vs Exponential Moving Average')
```

# **Individual Graph**

```python
colors = ['blue','orange','yellow','red']
i=0
for column in df.columns[1:]:
    plt.figure(figsize=(10,5))
    plt.plot(df['Date'], df[column], label=column,color=colors[i])
    plt.title(column)
    i += 1
    plt.legend()
    plt.show()
```

```python


colors = ['blue','orange','yellow','red']
i=0
for column in df.columns[1:]:
    plt.figure(figsize=(10,5))
    plt.hist2d(df['Date'], df[column],bins=20, label=column,color=colors[i],cmap='summer',norm=cls.LogNorm())
    plt.title(column)
    i += 1
plt.legend()
plt.show()
```
