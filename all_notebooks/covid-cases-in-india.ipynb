```python
death=pd.read_csv("/kaggle/input/india-covid19-cases-till-date/India_covid19_deaths_data.csv")
```

```python
confirmed_cases=pd.read_csv("/kaggle/input/india-covid19-cases-till-date/India_covid19_confirmed_cases_data.csv")
```

```python
death.head(5)
```

```python
confirmed_cases.head()
```

```python
df=pd.concat([confirmed_cases,death],axis=0)
```

```python
df.head(5)
```

```python
df.drop(["Province/State","Lat","Long","Country/Region"],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
dict_0={1:"confirmed_cases",0:"death"}
```

```python
df.head
```

```python
df["impact"]=df["1/30/20"].map(dict_0)
```

```python
df["impact"].head()
```

```python
df.head(5)
```

```python
df_t=df.T
```

```python
df_t.head()
```

```python
df_t.index.name="Date"
```

```python
df_t.columns=["Confirmed_cases","Death"]
```

```python
df1=df_t.iloc[:-1]
```

```python
df2=df1.reset_index()
```

```python
df2.head()
```

```python
from datetime import datetime
df2["Date"]=pd.to_datetime(df2["Date"],dayfirst=True)
```

```python
df2.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly
import plotly.express as px
import plotly.graph_objects as go
import cufflinks as cf
#import plotly.offline as pyo
#from plotly.offline import init_notebook
#_mode,plot,iplot
import folium
from folium import plugins
plt.rcParams["figure.figsize"]=10,12
import warnings
warnings.filterwarnings("ignore")
fig=go.Figure()
fig.add_trace(go.Scatter(x=df2["Date"],y=df2["Confirmed_cases"],mode="lines+markers",name="Total Cases"))
fig.update_layout(title_text="trend of confirmed cases",plot_bgcolor="rgb(230,230,230)")
fig.show()
```

```python
import plotly.express as px
fig=px.bar(df2,x="Date",y="Confirmed_cases",barmode="group",height=400)
fig.update_layout(title_text="Coronavirus daily cases in india",plot_bgcolor="gold")
fig.show()
```
