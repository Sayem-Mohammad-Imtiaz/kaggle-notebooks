```python
# import libraries
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
import pandas as pd 
```

```python
# read file
df = pd.read_csv('/kaggle/input/churn-modeling-dataset/Churn_Modelling.csv')
df
```

```python
# select features to train
# we don't use RowNumber, CustomerId and Surname columns because it don't have meaning

X = df.iloc[:, 3:-1]
X
```

```python
# define target column
y = df['Exited']
y
```

```python
# define category columns in features
cat_features = ['Geography', 'Gender']
```

```python
# split X, y for train and test sections
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=77)
```

```python
# define model: this case we use default parameter
model = CatBoostClassifier()
```

```python
# train model
model.fit(X_train, y_train, cat_features)
```

```python
# model prediction
y_pred = model.predict(X_test)
y_pred
```

```python
# model evaluation
print(classification_report(y_pred, y_test))
```
