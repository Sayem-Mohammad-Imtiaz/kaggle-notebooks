Seeing the data, class.csv contains the details of the class like mammals, amphibians, etc. alongwith number of animals in this class, its code and the names of the animals

animal.csv contains the details of the animals in a 1/0 format suggesting if a particular feature like hair / eggs / milk, etc. is present or not alongwith the class type - we actually only need animal.csv to make a prediction

```python
#Classifiying data and target
X = df2.iloc[:,1:17].values   # Not considering the name of the animal - placing it 
y = df2.iloc[:,17].values     # Class number to be assigned (labels)

# Separating into test and training data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

```python
#Standardizing data

from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(X_train)
```

**Different Models**
----------------

## SVM ##

Support Vector Machine is a classifier which separates the data set using a hyper plane. It selects the optimal hyper plane which maximizes the distance between the two data set. It uses a pre-formatted labelled data to give an output.![enter image description here][1]


  [1]: https://qph.ec.quoracdn.net/main-qimg-675fedee717331e478ecfcc40e2e4d38

```python
from sklearn.metrics import accuracy_score
from sklearn import svm
model = svm.LinearSVC()
model.fit(X_train, y_train)

# make predictions
expected = y_test
predicted = model.predict(X_test)

print ('The accurancy of SVM  is ' + str(accuracy_score(expected, predicted)))
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
```

## Decision Trees ##
The decision tree classifiers organizes a series of conditions (features) in a tree like structure. In a decision tree the leaf nodes are the final classification labels where as the non leaf nodes are test conditions and branch nodes are the output on the test conditions.![enter image description here][1]



  [1]: http://www.cse.unsw.edu.au/~billw/cs9414/notes/ml/06prop/id3/dtree.gif

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X_train,y_train)

# make predictions
expected = y_test
predicted = model.predict(X_test)

print ('The accurancy of Decision Tree Classifier  is ' + str(accuracy_score(expected, predicted)))
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
```

**Linear Discriminant Analysis**
----------------------------
It is a method of finding a linear combination of variables which best separates two or more classes. LDA is not a classification algorithm, although it makes use of class labels. However, the LDA result is mostly used as part of a linear classifier.

```python
from sklearn.metrics import accuracy_score
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
model = LinearDiscriminantAnalysis()

model.fit(X_train,y_train)

# make predictions
expected = y_test
predicted = model.predict(X_test)

print ('The accurancy of Linear Discriminant Analysis  is ' + str(accuracy_score(expected, predicted)))
# summarize the fit of the model
print(metrics.classification_report(expected, predicted))
```
