```python
import os
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.image as mpimg
```

##  Handy routine to print messages in RGB color

```python
def print_in_color(txt_msg,fore_tupple,back_tupple,):
    #prints the text_msg in the foreground color specified by fore_tupple with the background specified by back_tupple 
    #text_msg is the text, fore_tupple is foregroud color tupple (r,g,b), back_tupple is background tupple (r,g,b)
    rf,gf,bf=fore_tupple
    rb,gb,bb=back_tupple
    msg='{0}' + txt_msg
    mat='\33[38;2;' + str(rf) +';' + str(gf) + ';' + str(bf) + ';48;2;' + str(rb) + ';' +str(gb) + ';' + str(bb) +'m' 
    print(msg .format(mat))
```

# Generate a listing of class names and train, test amd validation file counts

```python
train_dir='../input/gender-classification-from-an-image/train'
dir_list=os.listdir(train_dir)
msg='{0:8s}{1:4s}{2:^35s}{1:4s}{3:11s}{1:3s}{4:10s}{1:3s}{5:11s}{6}'
msg=msg.format('Class Id', ' ', 'class name', 'Train Files','Test Files', 'Valid Files','\n')
print_in_color(msg, (255,0,0), (255,255,255))
klass_list= sorted(os.listdir(train_dir))
for i, klass in enumerate (klass_list):
    file_path=os.path.join(train_dir,klass)
    train_files_list=os.listdir(file_path)
    train_file_count=str(len(train_files_list))
    msg='{0:^8s}{1:4s}{2:^35s}{1:4s}{3:^11s}{1:3s}{4:^10s}{1:3s}{5:^11s}'
    msg=msg.format(str(i), ' ',klass, train_file_count,'5', '5')
    toggle=i% 2   
    if toggle==0:
        back_color=(255,255,255)
    else:
        back_color=(191, 239, 242)
    print_in_color(msg, (0,0,0), back_color)
```

# This code plots dample images from each class

```python
test_dir='../input/gender-classification-from-an-image/test'
classes=len(os.listdir(test_dir))
fig = plt.figure(figsize=(16,20))
for i, klass in enumerate(klass_list):
    klass_path=os.path.join(test_dir, klass)
    f_list=os.listdir(klass_path)
    if i <2:
        row= i
        for column in range(5): 
            j=row * 5 + column
            f_path=os.path.join(klass_path, f_list[column])
            img = mpimg.imread(f_path)
            a = fig.add_subplot(row +1, 5, column+1)
            a.axis("off")
            t_split=str(f_path).split(r'/')
            title=t_split[-2] +' ', t_split[-1]
            a.set_title(title, color='g', fontsize=16)
            plt.imshow(img)
    else:
        break
```
