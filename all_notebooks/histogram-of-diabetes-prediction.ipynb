```python
# to display the histogram of any column use this code along with their size and color.
# eg: histogram of Pregnancies column

import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("../input/diabetes-prediction/Diabetespred.csv")
df['Pregnancies'].hist(figsize=(10,8),color='red')
plt.title('Records of pregnancies')
plt.xlabel('Number')
plt.ylabel('Limit')
```
