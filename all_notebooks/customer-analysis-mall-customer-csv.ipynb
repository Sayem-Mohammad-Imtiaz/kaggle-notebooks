```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/kaggle/input/mall-customers/Mall_Customers.csv')
df.head()
```

```python
df.shape
```

```python
sns.countplot(df['Genre'])
```

```python
plt.scatter(df['Age'],df['Spending Score (1-100)'])
```

```python
plt.scatter(df['Annual Income (k$)'],df['Spending Score (1-100)'])
```

```python
x=df[['Genre','Annual Income (k$)','Age']]
x['Genre']=pd.get_dummies(x['Genre'],drop_first=True)
x
```

```python
score=[]
for i in range(1,10):
    model=KMeans(n_clusters=i)
    model.fit(x)
    score.append(model.inertia_)
score
```

```python
plt.plot(range(1,10),score)
```

```python
df['Genre']=pd.get_dummies(x['Genre'],drop_first=True)
```

```python
kn= KMeans(n_clusters=5,init = 'k-means++', max_iter = 300, n_init = 10)
y_pred=kn.fit_predict(df.drop(['CustomerID'],axis=1))
df['cluster']=y_pred
df
```

```python
df['cluster']=y_pred
df1=df[df.cluster==0]
df2=df[df.cluster==1]
df3=df[df.cluster==2]
df4=df[df.cluster==3]
df5=df[df.cluster==4]

plt.scatter(df1['Annual Income (k$)'],df1['Spending Score (1-100)'],color='red')
plt.scatter(df2['Annual Income (k$)'],df2['Spending Score (1-100)'],color='green')
plt.scatter(df3['Annual Income (k$)'],df3['Spending Score (1-100)'],color='yellow')
plt.scatter(df4['Annual Income (k$)'],df4['Spending Score (1-100)'],color='blue')
plt.scatter(df5['Annual Income (k$)'],df5['Spending Score (1-100)'],color='black')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')

plt.scatter(kn.cluster_centers_[:,2],kn.cluster_centers_[:,3],color='black',marker='*',label='centers')

plt.legend()
```
