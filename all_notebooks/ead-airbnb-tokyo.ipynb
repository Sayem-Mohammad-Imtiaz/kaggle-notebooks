```python
def get_disponiveis(ano_base):
    disponiveis = []
    df_available = df_calendar[df_calendar['available'] == 't'] 
    for i, row in df_available.iterrows():
        data = dt.datetime.strptime(str(row['date']), '%Y-%m-%d').date()
        if (data.year == ano_base):
            preco = float(row['price'].replace('$', '').replace(',', ''))
            preco_ajustado = float(row['adjusted_price'].replace('$', '').replace(',', ''))
            disponiveis.append([row[0], data, row['available'], preco, preco_ajustado])
    return disponiveis
```

```python
disp = get_disponiveis(2021)
print('ok')
```

```python
#max(map(lambda x: x[3], disp))
precos_mes = {}
precos_medios_mes = {}
total_acomodacoes_mes = {}
for mes in mapa_meses:
    precos_mes[mes] = 0
    total_acomodacoes_mes[mes] = 0
for item in disp:
    precos_mes[item[1].month] += item[3]
    if item[3] != 0:
        total_acomodacoes_mes[item[1].month] += 1
for mes in mapa_meses:
    if total_acomodacoes_mes[mes] == 0:
        precos_medios_mes[mes] = 0
    else:
        precos_medios_mes[mes] = precos_mes[mes] / total_acomodacoes_mes[mes]
    print(mapa_meses[mes], precos_medios_mes[mes])
```

```python
m = []
v = []
for key, mes in mapa_meses.items():
    m.append(mes)
for key, valor in precos_medios_mes.items():
    v.append(valor)
    
plt.title("Média de preços")
plt.ylabel("$")
plt.bar(m, v)
```
