# First, I'll import the dataset into Python using pandas:

```python
data = pd.read_csv('/kaggle/input/groundhog-day/archive.csv')
data.set_index('Year')
```

# Now, I am going to express the relationship between the average temperatures in Pennsylvania during February and March using graphs

```python
feb_av_temp = {}
mar_av_temp = {}
for row in data.values[:-1]:
    year = int(row[0])
    feb_av_temp[year] = row[5]
    mar_av_temp[year] = row[9]
#print(feb_av_temp,mar_av_temp)
fig = plt.figure(figsize=(12,5))
ax1 = fig.add_subplot()
ax1.set_title('Comparing Average Temperatures in Pennsylvania')
ax1.set_xlabel('Year')
ax1.set_ylabel('Temperature')
ax1.plot(feb_av_temp.keys(), feb_av_temp.values())
ax1.plot(mar_av_temp.keys(), mar_av_temp.values())
ax1.legend(['February Temperatures', 'March Temperatures'])
        
```

# I am going to filter my data so that we can see the average temperatures in Pennsylvania when Phil did not see his shadow.

```python
data_no_shadow = data.loc[(data['Punxsutawney Phil'] == 'No Shadow'),['Year','February Average Temperature (Pennsylvania)','March Average Temperature (Pennsylvania)']]
data_shadow = data.loc[(data['Punxsutawney Phil'] == 'Full Shadow'),['Year','February Average Temperature (Pennsylvania)','March Average Temperature (Pennsylvania)']]
fig = plt.figure(figsize=(12,4.8))
ax2 = fig.add_subplot()
ax2.set_title('Is Phil Accurate')
ax2.set_xlabel('year')
ax2.set_ylabel('temperature')
ax2.scatter(data_no_shadow.Year.astype(int),data_no_shadow['March Average Temperature (Pennsylvania)'])
ax2.scatter(data_shadow.Year.astype(int),data_shadow['March Average Temperature (Pennsylvania)'])
ax2.legend(['No Shadow', 'Full Shadow'])
```
