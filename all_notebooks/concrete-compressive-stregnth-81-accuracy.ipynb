#### Create train and test split

```python
Y=cs_df.pop('Concrete compressive strength(MPa, megapascals) ')
```

```python
scores={}
```

```python
X=cs_df
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, train_size=0.7, test_size = 0.3, random_state=100)
```

#### Buidling linear regression model

```python
from sklearn import datasets, linear_model, metrics
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
print('accuracy score: {}'.format(reg.score(X_test, y_test)))
```

```python
scores['Linear_test_score']=reg.score(X_test, y_test)
```

```python
print('accuracy score: {}'.format(reg.score(X_train, y_train)))
```

```python
scores['Linear_train_score']=reg.score(X_train, y_train)
```

#### Decision regressior

```python
from sklearn.tree import DecisionTreeRegressor 
  
# create a regressor object
regressor = DecisionTreeRegressor(random_state = 0) 
  
# fit the regressor with X and Y data
regressor.fit(X_train, y_train)
```

```python
print("test accuracy score for Decision regression",regressor.score(X_test,y_test))
```

```python
print("train accuracy score for Decision regression",regressor.score(X_train,y_train))
```

#### Tuning hyper parameters for Decision regressor

```python
parameters={"splitter":["best","random"],
            "max_depth" : [10,12,14,16,18,20],
           "min_samples_leaf":[1,2,3],
           
           "max_features":["auto","log2","sqrt"],
            }
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
tuning_model=GridSearchCV(regressor,param_grid=parameters,scoring='neg_mean_absolute_error',cv=3,verbose=3)
```

```python
tuning_model.fit(X_train,y_train)
```

```python
tuning_model.best_params_
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import KFold
#from sklearn.metrics import accuracy
```

#### Tuning hyper parameters for xgbhoost

```python
param_grid = {
        'max_depth': [3, 4, 5, 6],
       
        'gamma': [0.1,0.2,0.3],
        'learning_rate': [0.1,0.2,0.3],
        'subsample': [0.01,0.02,0.03]
        
}
```

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=10)
grid_search = RandomizedSearchCV(xgbhoost, param_grid,  cv=kfold)
grid_result = grid_search.fit(X_train,y_train)

print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_[ 'mean_test_score' ]
stds = grid_result.cv_results_[ 'std_test_score' ]
params = grid_result.cv_results_[ 'params' ]
```
