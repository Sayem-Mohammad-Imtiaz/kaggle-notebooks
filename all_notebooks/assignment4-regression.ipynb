# ** **Import Dataset** **

```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory
data_vw = pd.read_csv('/kaggle/input/used-car-dataset-ford-and-mercedes/vw.csv')
data_vw.head()
```

# ** **Data Transformation** **

```python
d = data_vw
d["age_of_car"] = 2020 - d["year"]
d = d.drop(columns = ["year"])
d["price_k"] = d["price"]/1000
d = d.drop(columns = ["price"])
d["mileage_k"] = d["mileage"]/1000
d = d.drop(columns = ["mileage"])


models = data_vw.model.unique()
model_mapper = {}

for i in range(len(models)):
   model_mapper[models[i]]=i

model = d["model"]
for i in range(len(model)):
    model[i]=model_mapper[model[i]]
models = d.model.unique()

transmissions = data_vw.transmission.unique()
transmissions_mapper = {}

for i in range(len(transmissions)):
   transmissions_mapper[transmissions[i]]=i

transmission = d["transmission"]
for i in range(len(transmission)):
    transmission[i]=transmissions_mapper[transmission[i]]
transmissions = d.transmission.unique()

fuelTypes = data_vw.fuelType.unique()
fuelTypes_mapper = {}

for i in range(len(fuelTypes)):
   fuelTypes_mapper[fuelTypes[i]]=i

fuelType = d["fuelType"]
for i in range(len(fuelType)):
    fuelType[i]=fuelTypes_mapper[fuelType[i]]
fuelTypes = d.fuelType.unique()
d.sample(10)
```

# ** **Simple linear model ( use any attribute)** **

```python
mileage = d.mileage_k
price = d.price_k

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(mileage,price, test_size=0.3 )

mileage = np.array(X_train).reshape((-1, 1))

price = np.array(y_train)

print(mileage,price)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(mileage, price)
print(model.coef_,'* X + ', model.intercept_)
test= np.array(X_test).reshape((-1, 1))
y_t= np.array(y_test).reshape((-1, 1))
y_pred = model.predict(test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(mileage, price, color='black')
plt.plot(test, y_pred, color='blue', linewidth=3)
plt.xticks(())
plt.yticks(())
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
from sklearn import metrics
import sklearn.metrics as sm
import math
print('Intercepts: ', model.intercept_)
print('Coefficients: ', model.coef_)
print("Mean absolute error =", round(sm.mean_absolute_error(test, y_pred), 2))
print("RMSE = " ,math.sqrt(round(sm.mean_squared_error(test, y_pred), 2)))
print("Mean squared error =", round(sm.mean_squared_error(test, y_pred), 2)) 
print("Median absolute error =", round(sm.median_absolute_error(test, y_pred), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(test, y_pred), 2)) 
print("R2 score =", round(sm.r2_score(test, y_pred), 2))
```

```python
mileage = d.mileage_k
price = d.price_k

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(mileage,price, test_size=0.3 )

```

```python
#!/usr/bin/env python
from pylab import *
from scipy.optimize import curve_fit

x = np.array(X_train)
y = np.array(y_train)
def func(x, a, c, d):
    return a*np.exp(c*x)+d
```

```python
popt, pcov = curve_fit(func, x, y, p0=(1, 1e-6, 1))
print(popt)
print(pcov)
```

```python
xx = np.linspace(0, 200, 70)
yy = func(xx, *popt)
plot(x, y, 'ko') 
plot(xx, yy)
X_test=np.array(X_test)
y_pred = func(X_test, *popt)
print (func(0, *popt))
print (y_pred)
```

```python
price = np.array(y_test)
import sklearn.metrics as sm
from sklearn.metrics import mean_squared_error
print('Intercepts: ',func(0, *popt ))
print('Coefficients: ', popt)
print("Mean squared error: %.2f" % mean_squared_error(price, y_pred))
      
print("Mean absolute error =", round(sm.mean_absolute_error(price, y_pred), 2)) 
print("Mean squared error =", round(sm.mean_squared_error(price, y_pred), 2)) 
print("RMSE = " ,math.sqrt(round(sm.mean_squared_error(price, y_pred), 2)))
print("Median absolute error =", round(sm.median_absolute_error(price, y_pred), 2)) 
print("Explain variance score =", round(sm.explained_variance_score(price, y_pred), 2)) 
print("R2 score =", round(sm.r2_score(price, y_pred), 2))
```
