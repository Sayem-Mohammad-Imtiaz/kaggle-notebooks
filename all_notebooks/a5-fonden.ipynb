* A breakdown of Covid cases and deaths per continent

CSV file pulled from this source: https://ourworldindata.org/coronavirus-source-data

```python
#Opening the CSV file and printing the head

df = pd.read_csv("../input/whocoviddatacsv/owid-covid-data.csv")
df.head()
```

```python
#Cleaning up the data, and seeing the most recent cases to date

df = df[["continent", "location", "date", "total_cases", "new_cases", "new_deaths", "gdp_per_capita", "life_expectancy"]]
df.tail(10)
```

# Total Number of New Cases
How many total cases have been recorded since January 2020?

```python
print(df['new_cases'].sum())
```

# Jan-Present Covid Trend
The scatter plot below shows the trend in Total Covid cases over the world from Jan 2021 to date.

```python
import plotly.express as px

#Remove the null values
df = df[df.continent.notnull()]

#Create line chart
fig = px.line(df, x="date", y="new_cases", color="continent", line_group="continent", hover_name="continent",
        line_shape="spline", render_mode="svg", labels={'new_cases':'# of Covid Cases', 'continent': 'Continent', 'date': 'Timeline by Month'})
fig.show()
```

# Cases Per Continent
Finding out what the percentages are per country, using a pie chart.

```python
#Create pie chart
fig = px.pie(df, values='new_cases', names='continent', title='Total Cases of Covid Per Continent')
fig.show()

#Print the sums to double check it stats are correct
print(df.sum()['new_cases'])
```

```python
#Total Number of Deaths

#Create pie chart
fig = px.pie(df, values='new_deaths', names='continent', title='Total Number of Covid Deaths Per Continent')
fig.show()

#Print the sums to double check it stats are correct
print(df.sum()['new_deaths'])
```

# United States Trend
The chart below looks at the US specifically, showing # of new covid cases over time, against the # of deaths.

```python
#Isolate the United States
united_states = df[df.location == 'United States']

#Create Bar Chart
fig = px.bar(united_states, x='date', y='new_cases',
             hover_data=['date', 'new_cases'], color='new_deaths',
             labels={'new_cases':'# of US Covid Cases', 'new_deaths': '# of Deaths', 'date': 'Timeline by Month'}, height=400)
fig.show()
```
