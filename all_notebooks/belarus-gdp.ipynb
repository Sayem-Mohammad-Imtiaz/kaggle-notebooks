```python
gdp = pd.read_csv('../input/gdp-per-capita-all-countries/GDP.csv')
gdp = gdp.rename(columns={'Country ': 'Country'})
df = gdp

# leave only country & years columns
gdp = gdp[['Country'] + [str(i) for i in range(1990, 2020)]]

# filter countries of interest
countries = ['Belarus', 'Lithuania', 'Latvia', 'Poland', 'Germany', 'Russian Federation', 'Ukraine',
             'Thailand',
             'United States']
gdp = gdp[gdp['Country'].isin(countries)]

# transpose & use first row as columns names
gdp = gdp.transpose().rename(columns=gdp.transpose().iloc[0])
gdp = gdp.drop(gdp.index[0])
```

```python
# ВВП на душу населения (далее просто ВВП)
gdp.plot(figsize=(15,7))
```

```python
# изменение ВВП относительно прошлого года
((gdp - gdp.shift(1)) / gdp).plot(figsize=(15,7)).grid(True)
```

```python
# изменение ВВП относительно прошлого года, скользящее среднее на 4 года
diff = ((gdp - gdp.shift(1)) / gdp)
diff.rolling(window=4).mean().plot(figsize=(15,7)).grid(True)
```
