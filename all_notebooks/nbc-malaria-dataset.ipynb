```python
import pandas as pd
import numpy as np
import math
import os

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

from skimage import io
from skimage import feature

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

import cv2
```

```python
#LOAD IMAGE DATASET
terinfeksi_path = '../input/malaria-dataset/terinfeksi'
tidak_terinfeksi_path = '../input/malaria-dataset/tidak_terinfeksi'

terinfeksi = os.listdir(terinfeksi_path)
tidak_terinfeksi = os.listdir(tidak_terinfeksi_path)

print('Done')
```

```python
#VISUALISASI
plt.figure(figsize = (8,8))
for i in range(3):
    plt.subplot(1, 3, i+1)
    img = cv2.imread(terinfeksi_path + "/" + terinfeksi[i])
    plt.imshow(img)
    plt.title('Terinfeksi actual')
    plt.tight_layout()

plt.figure(figsize = (8,8))
for i in range(3):
    plt.subplot(1, 3, i+1)
    img = cv2.imread(terinfeksi_path + "/" + terinfeksi[i])
    edges = cv2.Canny(img,25,255,L2gradient=False)
    plt.imshow(edges,cmap='gray')
    plt.title('Seg. canny')
    plt.tight_layout()
    
plt.figure(figsize = (8,8))
for i in range(3):
    plt.subplot(1, 3, i+1)
    img = cv2.imread(tidak_terinfeksi_path + "/" + tidak_terinfeksi[i])
    plt.imshow(img)
    plt.title('Tidak Terinfeksi actual')
    plt.tight_layout()

plt.figure(figsize = (8,8))
for i in range(3):
    plt.subplot(1, 3, i+1)
    img = cv2.imread(tidak_terinfeksi_path + "/" + tidak_terinfeksi[i])
    edges = cv2.Canny(img,25,255,L2gradient=False)
    plt.imshow(edges,cmap='gray')
    plt.title('Seg. Canny')
    plt.tight_layout()

    

plt.show()
```

```python
#CANNY DAN HUMOMENT
x= 0
x = np.array([['h1','h2','h3','h4','h5','h6','h7']])

for i in range(len(terinfeksi)):
    img = cv2.imread(terinfeksi_path + "/" + terinfeksi[i])
    edges = cv2.Canny(img,25,100)
    a = cv2.HuMoments(cv2.moments(edges)).flatten()
    x = np.vstack((x,a))

for i in range(len(tidak_terinfeksi)):
    img = cv2.imread(tidak_terinfeksi_path + "/" + tidak_terinfeksi[i])
    edges = cv2.Canny(img,25,100)
    a = cv2.HuMoments(cv2.moments(edges)).flatten()
    x = np.vstack((x,a))

print('Done')
```
