# Crime Agaisnt Women

```python
sns.set_style('darkgrid')
sns.set_palette('coolwarm')
a=[]
for i in df['Crime against women(Act)']:
    if str(i)=='nan':
        continue
    if ',' in str(i):
        p=str(i).split(',')
        for x in p:
            a.append(int(x))
    else:
        a.append(int(i))
a.sort()
fig = plt.figure(figsize=(20,7.5))
sns.countplot(x=a[:272])
plt.title('Crime Against Women In IPC(Act)')
plt.xlabel('IPC(Act)')
print('''
       Criminal Acts
       
       1: 'Murder with Rape',
       2: 'Dowry Deaths', 
       3: 'Suicide',
       4: 'Kidnapping',
       5: 'Acid Attack', 
       6: 'Cruelty by Husband/in-laws',
       7: 'Rape only',
       8: 'Assault on Women with Intent to Outrage her Modesty',
       9: 'Cyber Crimes against Women' ,
      10: 'Protection of Children from Sexual Offences Act'.
''')
```

# Citywise Kidnapping Report

```python
A = df[df['City ']=='Ghaziabad']
a1 = A['Kidnnaping:\nNumber of child victims'].sum()
b1 = A['Number of male(adult) '].sum()
c1= A['Number of female(adult)'].sum()

B = df[df['City ']=='Lucknow']
a2 = B['Kidnnaping:\nNumber of child victims'].sum()
b2 = B['Number of male(adult) '].sum()
c2 = B['Number of female(adult)'].sum()

C = df[df['City ']=='Kanpur']
a3 = C['Kidnnaping:\nNumber of child victims'].sum()
b3 = C['Number of male(adult) '].sum()
c3 = C['Number of female(adult)'].sum()
a = [a1,a2,a3]
b = [b1,b2,b3]
c = [c1,c2,c3]

City=['Ghaziabad','Lucknow','Kanpur']
plt.figure(figsize=(20,8))
sns.set_style('white')
sns.set_palette('coolwarm')
x=np.arange(len(City))
plt.title("City wise no of victims")
plt.xticks(x,City,rotation='horizontal')
plt.bar(x-0.2,a,width=0.2,label="Child ",color='#ffa600')
plt.bar(x,b,width=0.2,label="Male",color='#ff6361')
plt.bar(x+0.2,c,width=0.2,label="Female",color='#bc5090')
plt.xlabel("City")
plt.title("City Wise No of Kidnappings")
plt.ylabel("No of Cases")
plt.legend()
plt.grid()
plt.show()
```

1. >               In Ghaziabad Children And Male Cases Of Kidnappings Are High.
2. >               Female Cases Of Kidnappings are Very Low In Lucknow But Child Cases have large No's. 
3. >               Kanpur Has Low Kidnappings in All the Categories.

# Citywise Crime Report

```python
A = df[df['City ']=='Ghaziabad']
a1 = A['Number of child victims'].sum()
b1 = A['Number of male victims(adult)'].sum()
c1= A['Number of female victims(adult)'].sum()

B = df[df['City ']=='Lucknow']
a2 = B['Number of child victims'].sum()
b2 = B['Number of male victims(adult)'].sum()
c2 = B['Number of female victims(adult)'].sum()

C = df[df['City ']=='Kanpur']
a3 = C['Number of child victims'].sum()
b3 = C['Number of male victims(adult)'].sum()
c3 = C['Number of female victims(adult)'].sum()
a = [a1,a2,a3]
b = [b1,b2,b3]
c = [c1,c2,c3]

City=['Ghaziabad','Lucknow','Kanpur']
plt.figure(figsize=(18,10))
sns.set_style('white')
sns.set_palette('coolwarm')
x=np.arange(len(City))
plt.title("City wise no of victims")
plt.xticks(x,City,rotation='horizontal')
plt.bar(x-0.2,a,width=0.2,label="Child ",color='#ffa600')
plt.bar(x,b,width=0.2,label="Male",color='#ff6361')
plt.bar(x+0.2,c,width=0.2,label="Female",color='#bc5090')
plt.xlabel("City")
plt.ylabel("No of victims")
plt.legend()
plt.grid()
plt.show()
```

1. > Lucknow has the highest no of male victims and lowest no of child victims.
2. >   Kanpur and Ghaziabad has equal no of child victims
