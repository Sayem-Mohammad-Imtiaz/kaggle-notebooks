```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../input/notcias-publicadas-no-brasil/Historico_de_materias.csv", index_col=0)
```

```python
df.head()
```

```python
print("Quantidade de registros {} . \n".format(df.shape[0]))

print("Existem {} tipos de cadernos como  {}... \n".format(len(df.assunto.unique()),
                                                                           ", ".join(df.assunto.unique()[0:5])))
```

```python
assuntos = df.groupby("assunto")

# Summary statistic of all countries
assuntos.describe().head()
```

```python
plt.figure(figsize=(15,10))
assuntos.size().sort_values(ascending=False).plot.bar()
plt.xticks(rotation=50)
plt.xlabel("Cadernos")
plt.ylabel("Quantidade de Mat√©rias")
plt.show()
```

```python
titulo = df.titulo[0] + df.titulo[1] + df.titulo[2] + df.titulo[3] + df.titulo[4] 

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(titulo)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(titulo)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
