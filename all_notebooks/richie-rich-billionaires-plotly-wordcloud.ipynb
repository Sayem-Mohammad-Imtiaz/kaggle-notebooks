```python
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename=='Billionaire.csv':
            path = os.path.join(dirname, filename)
        else:
            pass
```

```python
import numpy as np
import pandas as pd

from PIL import Image

from wordcloud import WordCloud
from matplotlib import pyplot as plt

df= pd.read_csv(path)

import requests
import io

response = requests.get("https://raw.githubusercontent.com/ambader/ambader/main/MISC/cash_large.png")
cash_mask = np.array(Image.open(io.BytesIO(response.content)))

df.NetWorth = df.NetWorth.str.replace("$","",regex=True).str.replace(" B","",regex=True).values.astype(float)

dx= df[df["Rank"]<2000]
dd=pd.DataFrame(index=dx.Name)
dd["nw"]=dx.NetWorth.values
d=dd.to_dict()['nw']

plt.figure(figsize= (10,25))
wordcloud = WordCloud(colormap="YlGn",width = 200, height = 500, background_color ="honeydew",min_font_size = 1, mask=cash_mask,max_words=2000, prefer_horizontal=0.5).generate_from_frequencies(frequencies=d)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
