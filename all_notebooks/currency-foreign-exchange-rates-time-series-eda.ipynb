# **<span style="color:blue">EDA </span> of closing value of Currency1 wrt Currency2**

### **<span style="color:red"> Slug </span> column is of form "ABC/PQR"**
#### where ABC and PQR are standard active currency codes as of ISO 4217. 

### **<span style="color:red">Currency </span> contains a currency code**
#### which is the unit of the value in the tables. 

#### For example, if slug is JPY/INR and currency is INR it means that and <span style="color:blue"> x </span> is in INR - 
## <span style="color:blue"> 1 JPY = x INR </span> (Cur1 = x Cur2)

```python
### OPTIONS available for Cur1
for x in set(df['slug'].str.split('/', expand=True)[0].unique()):
    print (x)
```

## **<span style="color:red">INR (Indian Rupee)</span>**

```python
def scatter_plot(df_IC):
    for Cur2 in df_IC['currency'].unique():
        plt.figure(figsize=(18,6))
        df_temp = df_IC[df_IC['currency'] == Cur2]
        plt.scatter(df_temp['date'], df_temp['close'])
        plt.title("1 " + Cur1 + " = x " + Cur2, fontsize = 20,fontweight = 'bold', color='blue')
        plt.xlabel("Date", fontsize = 14, fontweight='bold', color='red')
        plt.ylabel('Closing Value', fontsize = 14, fontweight='bold', color='red')
        plt.show()
```

```python
Cur1 = 'INR'
df_IC = df[df['slug'].str.contains(Cur1 + '/')]
scatter_plot(df_IC)
```
