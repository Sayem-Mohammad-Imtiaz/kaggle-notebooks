```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB,MultinomialNB
from sklearn.preprocessing import StandardScaler
```

```python
X = df.iloc[:,:3] # Feature Set
```

```python
X.head()
```

```python
y = df['size_category'] # Target variable
```

```python
y.head()
```

```python
sc =StandardScaler()
X = sc.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.20,random_state=1)
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
pred = model.predict(X_test)
print("Classification Report \n")
print(classification_report(y_test,pred))
```

```python
model = RandomForestClassifier(n_estimators=200)
model.fit(X_train,y_train)
pred = model.predict(X_test)
print("Classification Report \n")
print(classification_report(y_test,pred))
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
pred = model.predict(X_test)
print("Classification Report \n")
print(classification_report(y_test,pred))
```

```python
class_0_df = df[df['size_category']==0]
class_1_df = df[df['size_category']==1]
class_2_df = df[df['size_category']==2]
class_3_df = df[df['size_category']==3]
class_4_df = df[df['size_category']==4]
class_5_df = df[df['size_category']==5]
class_6_df = df[df['size_category']==6]
```

```python
col = [ class_0_df,
        class_1_df ,
        class_2_df ,
        class_3_df ,
        class_5_df ,
        class_6_df ,
]
```

```python
for name in col:
    print(name.shape)
```

```python
class_0_df_over = class_0_df.sample(29575,replace=True)
class_2_df_over = class_2_df.sample(29575,replace=True)
class_3_df_over = class_3_df.sample(29575,replace=True)
class_4_df_over = class_4_df.sample(29575,replace=True)
class_5_df_over = class_5_df.sample(29575,replace=True)
class_6_df_over = class_6_df.sample(29575,replace=True)
```

```python
col_2 = [class_0_df_over ,
                        class_1_df,
                        class_2_df_over ,
                        class_3_df_over ,
                        class_4_df_over ,
                        class_5_df_over ,
                        class_6_df_over ]
```

```python
for name in col_2:
    print(name.shape)
```

```python
balanced_df = pd.concat([class_0_df_over ,
                        class_1_df,
                        class_2_df_over ,
                        class_3_df_over ,
                        class_4_df_over ,
                        class_5_df_over ,
                        class_6_df_over ],
                        axis=0)
                        
```

```python
balanced_df['size_category'].value_counts()
```

```python
X = balanced_df.iloc[:,:3] # Feature Set
```

```python
X.head()
```

```python
y = balanced_df['size_category']
```

```python
sc =StandardScaler()
X = sc.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.20,random_state=1,stratify=y)
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
pred = model.predict(X_test)
print("Classification Report \n")
print(classification_report(y_test,pred))
```

```python
model = RandomForestClassifier(n_estimators=500)
model.fit(X_train,y_train)
pred = model.predict(X_test)
print("Classification Report \n")
print(classification_report(y_test,pred))
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
pred = model.predict(X_test)
print("Classification Report \n")
print(classification_report(y_test,pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=50)
model.fit(X_train,y_train)
pred = model.predict(X_test)
print("Classification Report \n")
print(classification_report(y_test,pred))
```
