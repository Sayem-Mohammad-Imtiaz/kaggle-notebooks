```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
# loading CSV file using pandas library 
df=pd.read_csv('../input/bank-note-authentication-uci-data/BankNote_Authentication.csv')
```

```python
# use to see how our data looks
df.head()
```

```python
# here we use iloc for seprating labels and features
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
# here we are dividing our tubles in train and test data so that we can evaluate how our model is working 
X_train,X_test,Y_train,Y_test=train_test_split(x,y,random_state=1,test_size=.3)

# here we are checking the shape of our train and test data
X_train.shape,X_test.shape
```

```python
# Now here we use to create create a model of neural network
# importing all nessucary library for modeling 
# for creating models using keras

import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import LambdaCallback
```

```python
# creating model
model=Sequential()

# model having input layer with four neurons and a hidden layer with ten neurons 

layer1=Dense(10, activation='relu',input_dim=4,kernel_regularizer=tf.keras.regularizers.l1(0.01))
model.add(layer1)

# model last layer having one neuron (binary output required)

model.add(Dense(1,activation='sigmoid',kernel_regularizer='l1'))

print_weights = LambdaCallback(on_epoch_end=lambda batch, logs: print(model.layers[0].get_weights()))

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])

history=model.fit(X_train,Y_train,batch_size=32,epochs=50,validation_data=(X_test,Y_test),use_multiprocessing=True,callbacks = [print_weights])
```

```python
# final weight 
model.layers[1].get_weights()
```

```python
Y_pred=model.predict(X_test)
y_pred=[]
for i in Y_pred:
    if i < .5:
        y_pred.append(0)
    else:
        y_pred.append(1)
y_pred
```

```python
# making confusion matrix to check for how much values our model confuse
from sklearn.metrics import confusion_matrix
confusion_matrix(Y_test,y_pred)
```

```python
# clssification report 
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```

```python
# ploating model accuracy
plt.plot(history.history['val_accuracy'])

plt.title("model accuracy")

plt.ylabel('accuracy')
plt.xlabel("epoch")
plt.legend(['train'],loc='upper left')

plt.show()
```

```python
# ploating model loss
plt.plot(history.history['val_loss'])

plt.title("model loss")

plt.ylabel('loss')
plt.xlabel("epoch")
plt.legend(['train'],loc='upper left')

plt.show()
```
