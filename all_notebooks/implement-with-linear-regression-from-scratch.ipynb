```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Reading data from CSV
dataset = pd.read_csv('/kaggle/input/salary/Salary.csv')
dataset.head(5)
```

```python
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
plt.scatter(x,y)
plt.plot(x,y)
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
```

### Separating Training Data and the test Data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

### Calculating Mean


```python
x_train_mean=x_train.mean()
y_train_mean=y_train.mean()
print(x_train_mean,y_train_mean)
```

```python
## Getting Coefficient values
num=0
den=0
for r in range(len(x_train)):
    num+=(x_train[r]-x_train_mean)*(y_train[r]-y_train_mean)
    den+=(x_train[r]-x_train_mean)**2
b1=num/den
b0=y_train_mean-b1*x_train_mean
print('Coefficients are:',float(b1),float(b0))
```

```python
## function to Predict values
def prediction(data):
    return b0+b1*data
```

```python
plt.scatter(x_train,y_train,color='green')
plt.plot(x_train, prediction(x_train),color = 'blue')
plt.title('training data vs predicted data on the training data')
plt.xlabel('Exp')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x_test,y_test,color='green')
plt.plot(x_test, prediction(x_test),color = 'blue')
plt.title('Test data vs predicted data on the training data')
plt.xlabel('Exp')
plt.ylabel('Salary')
plt.show()
```

### Model Evaluation Metrics for Regression

```python
from sklearn import metrics
```

```python
print(metrics.mean_absolute_error(y_test, prediction(x_test)))
print(metrics.mean_squared_error(y_test, prediction(x_test)))
print(np.sqrt(metrics.mean_squared_error(y_test, prediction(x_test))))
```
