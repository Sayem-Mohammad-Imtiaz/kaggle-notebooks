```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
myData= pd.read_csv("StudentsPerformance.csv")
cols = myData.columns
cols = cols.map(lambda x: x.replace(' ', '_'))
myData.columns = cols
#myData.info()
myData.head()


#myData.math_score.mean()
```

```python
# Quelles sont les moyennes des scores en math, reading et writing par genre
myData.groupby('gender').agg({'math_score':'mean', 'reading_score':'mean','writing_score':'mean'})
```

```python
# Quelles sont les moyennes des scores en math, reading et writing combinés par niveau d'éducation des parents? 

myData.groupby('parental_level_of_education').agg({'math_score':'mean', 'reading_score':'mean','writing_score':'mean'})
```

```python
#Quel est le pourcentage d'étudiants ayant réussi dans tous les tests (score>=50) par niveau de préparation au test
data=myData.query('math_score >=50 and reading_score>=50 and writing_score>=50' )
data.groupby('test_preparation_course').agg({'math_score':lambda x:np.mean(x)*100,'writing_score':lambda x:np.mean(x)*100,'reading_score':lambda x:np.mean(x)*100})

```
