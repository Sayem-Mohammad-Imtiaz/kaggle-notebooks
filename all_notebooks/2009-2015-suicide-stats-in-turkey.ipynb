Check the CSV File

```python
nRowsRead = 1000 # specify 'None' if want to read whole file
# who_suicide_statistics.csv may have more rows in reality, but we are only loading/previewing the first 1000 rows
data = pd.read_csv('/kaggle/input/suicide-rates-overview-1985-to-2016/master.csv')
nRow, nCol = data.shape
print(f'There are {nRow} rows and {nCol} columns')
```

Create data frame as given  columns with data in csv file
> We need this columns: 'country, year, sex, age, suicides_no' 
  So the codes are as follows

```python
dataFrame = data[['country','year','sex','age','suicides_no']] 
```

We can see data frame how look like in the following

```python
dataFrame.head()
```

Filtering data for Turkey

```python
# 1 - Filtering Pandas data frame
filteredData = dataFrame['country']=='Turkey'   # country == Turkey Output-> True or False 
dataFrame[filteredData]                         # Show True values in dataframe
```

dataFrame veriable is Filtered now

```python
dataFrame = dataFrame[filteredData] # Recreate dataFrame with true values in dataframe
```

Lets draw Age - Suicide Number Scatter Plot 

```python
dataFrame.plot(kind='scatter', x='age', y='suicides_no',color = 'red')
plt.xlabel('Age')              # label = name of label
plt.ylabel('Suicide Number')
plt.title('Age - Suicide Number Scatter Plot') 
plt.show()
```

Lets draw Sex - Suicide Number Scatter Plot 

```python
dataFrame.plot(kind='scatter', x='sex', y='suicides_no',color = 'red')
plt.xlabel('Sex')              # label = name of label
plt.ylabel('Suicide Number')
plt.title('Sex - Suicide Number Scatter Plot') 
plt.show()
```

Lets draw Year - Suicide Number Scatter Plot 

```python
dataFrame.plot(kind='scatter', x='year', y='suicides_no',alpha = 0.5,color = 'red')
plt.xlabel('Year')              # label = name of label
plt.ylabel('Suicide Number')
plt.title('Year - Suicide Number Scatter Plot') 
plt.show()
```
