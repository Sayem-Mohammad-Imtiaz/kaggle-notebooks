```python
#Question:What does the happiness rank have to do with the generosity element of a country?
data=df[['Happiness Rank','Generosity']]
data
```

```python
matrix=data.corr()
import seaborn as sns
sns.heatmap(matrix, annot=True)
plt.show()
```

```python
#very weak correlation between happiness Rank and generosity.Therefore, the happiness rank of the country doesn't have anything to do with the generosity score.
```

```python
#Does happiness score have something to do with the generosity score?
df1=df[['Happiness Score','Generosity']]
df1
```

```python
correlation1=df1.corr()
sns.heatmap(correlation1, annot=True)
plt.show()
```

```python
#low correlation score between Happiness Score and Generosity.
```
