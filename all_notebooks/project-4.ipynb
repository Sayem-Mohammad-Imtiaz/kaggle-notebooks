```python
df1 = pd.read_csv('/kaggle/input/frenchpresidentialelection2017/French_Presidential_Election_2017_First_Round.csv')
df1.head()
```

```python
df1=df1.drop(['Unnamed: 0','Department code','Constituency code','Commune code','Polling station','% Abs/Reg','% Vot/Reg','None of the above(NOTA)','% NOTA/Reg','% NOTA/Vot','% Nulls/Reg','% Nulls/Vot','% Exp/Reg','% Exp/Vot','First name','% Votes/Reg','% Votes/Exp','INSEE code','Polling station name','Postal code','Poll.St.-unique'],axis=1)
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
for col in df1.columns:
    if(df1[col].isnull().sum()!=0):
        if(df1.dtypes[col] == 'object'):
            df1[col].fillna(df1[col].value_counts().idxmax(), inplace = True)
        if(df1.dtypes[col] == 'int64'):
            df1[col].fillna(df1[col].mean(), inplace= True)
        if(df1.dtypes[col] == 'float64'):
            df1[col].fillna(df1[col].value_counts().idxmax(), inplace= True)
```

```python
for col in df1.columns:
    if(df1.dtypes[col] == 'object'):
        from sklearn.preprocessing import LabelEncoder
        label=LabelEncoder()
        df1[col]=label.fit_transform(df1[col])
```

```python
for col in df1.columns:
    if(df1.dtypes[col] == 'float64'):
        df1[col] = df1[col].astype(int)
```

```python
from sklearn.model_selection import train_test_split
train, test=train_test_split(df1, test_size=0.1, random_state=1)

def data_splitting(df1):
    x=df1.drop(['Voted'], axis=1)
    y=df1['Voted']
    return x, y


x_train, y_train=data_splitting(train)
x_test, y_test=data_splitting(test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression

log_model=LogisticRegression()
log_model.fit(x_train, y_train)
prediction=log_model.predict(x_test)
score= accuracy_score(y_test, prediction)
print(score)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
smote = XGBClassifier()
smote.fit(x_train, y_train)

# Predict on test
smote_pred = smote.predict(x_test)
accuracy = accuracy_score(y_test, smote_pred)
print("Test Accuracy is {:.2f}%".format(accuracy * 100.0))
```
