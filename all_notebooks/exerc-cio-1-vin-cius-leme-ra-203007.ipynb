### Questão 1 - Item C - Representação Gráfica
Para cada uma das variáveis, produza um ou mais gráficos, usando matplotlib, que descreva seu comportamento / caracteristica. Lembre-se que estes gráficos precisam ser compatíveis com a classificação da variável. 

```python
total_ele = df[['uf','total_eleitores']].groupby('uf',as_index = False).sum().sort_values(by = 'total_eleitores', ascending = False)
total_ele
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

altura = []
for i in total_ele['total_eleitores']:
   altura.append(i)
posicao = []
for i in range(0,26,1): 
   posicao.append(i)


fig, ax = plt.subplots(figsize=(20,10))
sns.barplot(x=total_ele['uf'], y=total_ele['total_eleitores'], ax=ax, data=total_ele,  palette='Blues_d')

ax.set_title('Total Eleitores x UF', fontdict={'fontsize':20})
ax.set_xlabel('UF', fontdict={'fontsize':20})
ax.set_ylabel('')
ax.set_ylabel('Total eleitores', fontdict={'fontsize':20})
ax.tick_params(labelsize=14)
ax.spines['bottom'].set_linewidth(2.5)

for axis in ['top', 'right', 'left']:
    ax.spines[axis].set_color(None)

for i in range(26):
   ax.text(x=posicao[i]-0.7, y=altura[i]+0.7, s=str(altura[i]), fontsize=14)

fig.tight_layout()
```

```python
total_ele_fem = df[['uf','gen_feminino']].groupby('uf',as_index = False).sum().sort_values(by = 'gen_feminino', ascending = False)
total_ele_fem
```

```python
grupos = total_ele_fem['uf']
valores = total_ele_fem['gen_feminino']
plt.title('Total de Eleitores Femininos x UF',fontdict={'fontsize':20})
plt.rcParams['figure.figsize'] = (20,10)
plt.bar(grupos, valores, color='orange')
plt.xlabel('UF')
plt.ylabel('Total de Eleitores femininos',fontdict={'fontsize':20})
plt.show()
```

```python
total_ele_masc = df[['uf','gen_masculino']].groupby('uf',as_index = False).sum().sort_values(by = 'gen_masculino', ascending = False)
total_ele_masc
```

```python
grupos = total_ele_masc['uf']
valores = total_ele_masc['gen_masculino']
plt.title('Total de Eleitores masculino x UF',fontdict={'fontsize':20})
plt.rcParams['figure.figsize'] = (20,10)
plt.bar(grupos, valores, color='green')
plt.xlabel('UF')
plt.ylabel('Total de Eleitores masculino',fontdict={'fontsize':20})
plt.show()
```

```python
total_ele_NI = df[['uf','gen_nao_informado']].groupby('uf',as_index = False).sum().sort_values(by = 'gen_nao_informado', ascending = False)
total_ele_NI
```

```python
grupos = total_ele_NI['uf']
valores = total_ele_NI['gen_nao_informado']
plt.title('Total de Eleitores gênero não informado x UF',fontdict={'fontsize':20})
plt.rcParams['figure.figsize'] = (20,10)
plt.bar(grupos, valores, color='grey')
plt.xlabel('UF')
plt.ylabel('Total de Eleitores gênero não informado',fontdict={'fontsize':20})
plt.show()
```

```python
valores_fem = total_ele_fem['gen_feminino']
valores_masc = total_ele_masc['gen_masculino']
valores_NI = total_ele_NI['gen_nao_informado']

x1 =  np.arange(len(valores_fem))
x2 = [x + 0.25 for x in x1]
x3 = [x + 0.25 for x in x2]

plt.bar(x1, valores_fem, width=0.25, label = 'Gênero Feminino', color = 'orange')
plt.bar(x2, valores_masc, width=0.25, label = 'Gênero Masculino', color = 'green')
plt.bar(x3, valores_NI, width=0.25, label = 'Gênero Não Informado', color = 'grey')

uf = total_ele['uf']
plt.xticks([x + 0.25 for x in range(len(valores_fem))], uf)

plt.legend()

plt.title("Total de eleitores gêneros Feminino x Masculino x Não Informado por estado", fontdict={'fontsize':20})
plt.show()
```

```python
gen_feminino = df['gen_feminino'].sum()
gen_masculino = df['gen_masculino'].sum()
gen_nao_informado = df['gen_nao_informado'].sum()

sum_gen = [gen_feminino,gen_masculino,gen_nao_informado]
legenda = ['Feminino','Masculino', 'Não Informado']
cores = ['#99b8c1','#e6ba3d','#2d2e3f']

fig, ax = plt.subplots(figsize=(10,8), subplot_kw=dict(aspect='equal'))

ax.pie(sum_gen, autopct='%0.1f%%', pctdistance=1.15, colors=cores)
ax.legend(labels=legenda, title='Gênero', loc='center left', bbox_to_anchor=(1.25,0,0.5,1))
ax.set_title('Porcetagem de eleitores por gênero',fontdict={'fontsize':20})
plt.show()
```
