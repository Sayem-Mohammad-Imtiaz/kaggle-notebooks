```python
import pandas as pd
df=pd.read_csv('/kaggle/input/the-human-freedom-index/hfi_cc_2020.csv')
df
```

```python
df.info()
df.dropna( inplace=True )
```

```python
cualitativos=list(df.select_dtypes(include = "object"))
cualitativos
```

```python
cuantitativos=list(df.select_dtypes(include = ['float64','int64']))
print(cuantitativos)
```

```python
rango=np.max(df.select_dtypes(include = ['float64','int64']))-np.min(df.select_dtypes(include = ['float64','int64']))
rango=pd.DataFrame(rango)
rango.index
rango=rango.rename(columns={0:'Rango'})
rango
```

```python
import scipy
from scipy import stats 
```

```python
moda=stats.mode(df.select_dtypes(include = ['float64','int64']))
moda1=pd.DataFrame(moda[0])
moda1.columns=cuantitativos
moda1
```

```python
moda1.T
```

```python
varianza=np.var(df.select_dtypes(include = ['float64','int64']))
varianza=pd.DataFrame(varianza)
varianza=varianza.rename(columns={0:'Varianza'})
varianza
```

```python

desviacion_estandar=np.std(df.select_dtypes(include = ['float64','int64']))
desviacion_estandar=pd.DataFrame(desviacion_estandar)
desviacion_estandar=desviacion_estandar.rename(columns={0:'Desviacion Estandar'})
desviacion_estandar
```

```python
datos = pd.concat([rango, varianza, desviacion_estandar], axis=1)
datos
```
