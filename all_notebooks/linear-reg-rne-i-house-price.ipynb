```python
import os 
import keras 
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
```

```python
df=pd.read_csv('/kaggle/input/bengaluru-house-price-data/Bengaluru_House_Data.csv')
df.head()
```

total_sqft, evin kaç metrekare büyüklüğünde olduğunu ifade ediyor. Kolaylık olsun diye "m2" olarak ismini değiştiriyorum. Ve type'ını öğreniyorum.

```python
df.rename(columns = {'total_sqft':'m2'},inplace=True)
df['m2'].dtypes
```

"m2" object type'ında olduğu için onu float'a çeviriyoruz.

```python
df['m2'] = pd.to_numeric(df['m2'], errors='coerce') 
```

```python
df.dtypes
```

m2 ve price arasındaki ilişkiye göre regression'ı oluşturacağımızdan bu değerlerdeki kaç tane null/NaN değeri olduğunu öğrenmemiz lazım.

```python
print(df.isnull().sum())
```

m2'de 247 tane NaN değer var. Ben bunları o sütunun ortalama değerleriyle doldurmak istiyorum.

```python
df["m2"].fillna(df["m2"].mean(), inplace=True)
```

Bu işlem sonrasında kaç tane NaN değeri kaldığına bakalım.

```python
print(df.isnull().sum())
```

Aşağıdaki gibi sns.pairplot kullanarak bütün feature'ların birbiriyle ilişkilerini de görsel bir şekilde detaylandırarak inceleyebiliriz.

```python
import seaborn as sns
sns.pairplot(df)
```

Heatmap ile de feature'ların arasındaki kolerasyonu görebiliriz. 

```python
sns.heatmap(df.corr()) 
```

```python
x = df[['m2']]
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(x_train,y_train)
```

Coefficient değeri öğrenelim.

```python
coeff_df = pd.DataFrame(lm.coef_,x.columns,columns=['Coefficient'])
coeff_df
```

```python
lm.score(x_test, y_test)
```

```python
#Grafik eksenlerini belirledik
df.plot(kind='scatter',x='m2',y='price', title='Housing Prices and Square Footage of Bengaluru')

#Fit kare bazında tahmini konut fiyatını bulduk.
y_pred = lm.predict(x) 

#Linear regression çizgisini çizdik.
plt.plot(x, y_pred, color='red')
```

Bu modelin error değerlerini öğrenelim.

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
