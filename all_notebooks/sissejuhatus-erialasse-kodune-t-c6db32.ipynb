Otsime välja pokemonid, kes on putukad ja on mürgised.

```python
putukad = df[df["Type 1"] == "Bug"]
mürgised_putukad = putukad[putukad["Type 2"] == "Poison"]
mürgised_putukad
```

Lisame mürgised putukad histogrammi.

```python
mürgised_putukad["Attack"].plot.hist()
```

Lisame kõik pokemonid "HP" ja "Defense" väärtuste järgi tihedusgraafikusse.

```python
df.plot.scatter("HP","Defense",alpha=0.2)
```

Grupeerime pokemonid "Type 1" parameetri järgi ning loetleme kokku, mitu "Legendary" pokemoni vastab igale  "Type 1"-le.

```python
df.groupby("Type 1")["Legendary"].count()
```
