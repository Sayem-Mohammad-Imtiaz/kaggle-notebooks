```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('/kaggle/input/euro-2020-top-players/euro2020.csv')
data.head()
```

```python
data.describe()
```

```python
data.info()
```

```python
most_goals = data[['Player','Goals']].sort_values(by="Goals",ascending=False)
most_goals = most_goals.head(10)
```

# Attacking Performance

# Most Goals

```python
plt.figure(figsize=(10,6))
sns.barplot(x=most_goals.Player,y=most_goals.Goals)
plt.xticks(rotation=45)
plt.title('Top 10 Most Goals Players')
```

```python
head_goals = data['Header goals'].sum() /data['Goals'].sum()
right_goals= data['Right foot goals'].sum() / data['Goals'].sum()
left_goals = data['Left foot goals'].sum() / data['Goals'].sum()
goal_dist =pd.DataFrame( {'Type':['Header', 'Right Foot','Left Foot'],
            'Rate':[head_goals,right_goals,left_goals],})
```

# Rate Of Goals With Header & Right Foot ,Left Foot

```python
plt.figure(figsize=(10,6))
sns.barplot(data=goal_dist,x='Rate',y='Type')
```

# The Most Effective Players

```python
data['MEP'] =data['Goals'] + data['Assists']
mes_player = data[['Player','MEP']].sort_values(by='MEP',ascending =False).head(5)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=mes_player,x='MEP',y='Player')
plt.xlabel("Most Effective Players")
```

# Success Shoot Rate

```python
data['Success Rate'] = data['Total attempts'] / data['Goals']
data['Success Rate'].replace(np.inf , 0 ,inplace=True)
most_success_rate =data[['Player','Success Rate']].sort_values(by='Success Rate',ascending=False).dropna().head(10)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=most_success_rate,x='Success Rate',y='Player')
plt.title("Most Success shoot Rates")
```

# Best  Score Performance Base On Time

```python
data['AVG PM'] = data['Minutes played'] / data['Goals']
data['AVG PM'].replace(np.inf ,np.nan ,inplace=True)
best_time_performance = data[['Player','AVG PM']].dropna().sort_values(by="AVG PM").head(10)
```

```python
plt.figure(figsize=(16,8))
sns.barplot(data=best_time_performance,x='Player',y='AVG PM')
plt.xticks(rotation =45)
plt.title("AVerage Goals Per Minute")
```

# Goals Scored By Positions

```python
 pos_goals =data[['Position','Goals']].groupby(['Position'],as_index=False).count().sort_values(by='Goals',ascending =False)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=pos_goals['Position'],y=pos_goals['Goals'])
```

# Most Goals Scored By Team

```python
country_goals = data[['Country','Goals']].groupby(['Country'],as_index=False).sum().sort_values(by='Goals',ascending=False)
```

```python
plt.figure(figsize=(12,6))
sns.pointplot(data=country_goals,x='Goals',y='Country')
plt.yticks(rotation=45)
```

# **Fastest Players **

```python
fastest_player = data[['Player','Top speed']].sort_values(by='Top speed',ascending=False).head(10)

plt.figure(figsize=(10,8))
sns.barplot(data=fastest_player,x='Top speed',y='Player')
fastest_player
```

# Defence Performance

```python
data['Defence Performance'] = data['Clearence attempted']+data['Balls recovered']+data['Blocks']+data['Tackels']+data['Shot blocks']
deff =data[['Player','Defence Performance']].dropna().sort_values(by='Defence Performance',ascending=False).head(10)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=deff,y='Player',x='Defence Performance')
```
