# Read data

```python
data_df = pd.read_csv("../input/first-pregnancy-age-in-europe/births_eu.csv")
```

# Clean the data

Transform `Value` field into a float.

```python
data_df["Value"] = data_df["Value"].apply(lambda x: str(x))
data_df["Value"] = data_df["Value"].apply(lambda x: x.replace(" ", ""))
data_df["Value"] = data_df["Value"].apply(lambda x: x.replace(",", ""))
data_df["Value"] = data_df["Value"].apply(lambda x: float(x))
```

# Filter the data

Get the teen data from overall data.  
Teen data will contains 2 age groups: 

* 'From 10 to 14 years'
* 'From 15 to 19 years'

# Top teen pregnancy per country and year

We exclude `European Union - 28 countries (2013-2020)` to focus on country level.

```python
teen_df = data_df.loc[data_df.AGE.isin(['From 10 to 14 years', 'From 15 to 19 years'])]
teen_df.loc[teen_df.GEO != "European Union - 28 countries (2013-2020)"].sort_values(by=["Value", "GEO", "TIME"], ascending=False).head(10)
```

Romania and Poland dominates the top, with mostly the transition years. Please note that Poland population is more than twice Romanian population.

# Top teen pregnancy per country

```python
tp_df = teen_df.loc[(teen_df.GEO != "European Union - 28 countries (2013-2020)")]
tp_df = tp_df.groupby(["GEO"])["Value"].sum().reset_index()
tp_df.sort_values(by=["Value"], ascending=False).head(10)
```

# Top infant pregnancy per country and year

Here we focus only on pre-teen pregnancies, i.e. `From 10 to 14 years` age interval. 

```python
teen_df.loc[(teen_df.AGE=='From 10 to 14 years') & (teen_df.GEO != "European Union - 28 countries (2013-2020)")].sort_values(by=["Value", "GEO", 'TIME'], ascending=False).head(10)
```

Romania dominates the top, with recent years showing an alarming increase in infant (pre-teens) pregnancy.

# Top infant pregnancy per country

```python
infant_df = teen_df.loc[(teen_df.AGE=='From 10 to 14 years') & (teen_df.GEO != "European Union - 28 countries (2013-2020)")]
infant_df = infant_df.groupby(["GEO"])["Value"].sum().reset_index()
infant_df.sort_values(by=["Value"], ascending=False).head(10)
```
