```python
df = pd.read_csv("../input/clicks-conversion-tracking/KAG_conversion_data.csv")
df.head()
```

```python
x=df.iloc[:, [6, 8]].values
x[0:5]
```

```python
from sklearn.cluster import Birch
```

```python
model = Birch(branching_factor = 100, n_clusters = 3, threshold = 0.6) 
  
# Fit the data (Training) 
model.fit(x) 
  
# Predict the same data 
pred = model.predict(x) 
  
# Creating a scatter plot 
plt.scatter(x[:, 0], x[:, 1],s=70, c = pred, cmap = 'rainbow', alpha = 0.7, edgecolors = 'b') 
plt.show()
```

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')
plt.scatter(x[pred == 0, 0], x[pred == 0, 1], s = 100, c = 'red', label = '916')
plt.scatter(x[pred == 1, 0], x[pred == 1, 1], s = 100, c = 'green', label = '936')
plt.scatter(x[pred == 2, 0], x[pred == 2, 1], s = 100, c = 'black', label = '1178')

#Plotting the centroids of the clusters
plt.show()
```
