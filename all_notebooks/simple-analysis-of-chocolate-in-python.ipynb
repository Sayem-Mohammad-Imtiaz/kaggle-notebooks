## **1: Import the necessary Python packages**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# use seaborn default plotting style
sns.set()
```

```python
# read the data
df = pd.read_csv('../input/flavors_of_cacao.csv')

# have a look
df.head()
```

## **2: Simple data cleaning**

```python
# remove some unnecessary strings `\n` and `\xa0`
df.columns = df.columns.str.replace('\n', ' ').str.replace('\xa0', '')

df.columns.tolist()
```

```python
# convert `string` to `float`
df['Cocoa Percent'] = df['Cocoa Percent'].apply(lambda row: row[:-1]).astype('float')

# check data types
df.dtypes
```

### **Question 1: Whatâ€™s the relationship between cocoa solids percentage and rating?**

```python
# look at correlation
df.corr()
```

The correlation between **Cocoa Percent** and **Rating** is about **0.1**, which shows that **Cocoa Percent** contributes little to the **Rating** of chocolate.

## **3: EDA**
### **Question 2: Where are the best cocoa beans grown?**

```python
# set marker size based on the cocoa percent of a chocolate
markerprop = {'s': .1*df['Cocoa Percent']**1.7, 'alpha': .4}

# scatterplot
ax = sns.lmplot(x='Broad Bean Origin', y='Rating', data=df,
                scatter_kws=markerprop, 
                fit_reg=False, legend=False, aspect=2.5)

# add averaged rating of each `Bean Origin`
df_copy = df.copy().groupby(['Broad Bean Origin'])['Rating'].mean()
plt.plot(df_copy, color='k', marker='o', alpha=.8, lw=3, label='Averaged Rating')

# rotate x tick label
ax.set_xticklabels(rotation=90)

# graph properties
plt.title('Rating of Chocoloate VS Bean Origin (2006-2017)', fontsize=20)
plt.xlabel('Broad Bean Origin', fontsize=15)
plt.ylabel('Rating (1-5)', fontsize=15)
plt.margins(.03)
plt.legend()

plt.show()
```

On average, most of the beans have rating above 3.

```python
# get the cocoa beans with the highest average rating
df_copy[df_copy == df_copy.max()]
```

We have **5** bean origin that produce the highest chocolate rating on average.

### **Question 3: Which countries produce the highest-rated bars?**

```python
# set marker size based on the cocoa percent of a chocolate
markerprop = {'s': .1*df['Cocoa Percent']**1.7, 'alpha': .2, 'color': 'navy'}

# scatterplot
ax = sns.lmplot(x='Company Location', y='Rating', data=df,
                scatter_kws=markerprop, 
                fit_reg=False, legend=True, aspect=2.5)

# add averaged rating
df_copy = df.copy().groupby(['Company Location'])['Rating'].mean()
plt.plot(df_copy, color='r', marker='o', alpha=.8, lw=3, label='Averaged Rating')

# # graph properties
ax.set_xticklabels(rotation=90)
plt.title('Rating of Chocoloate VS Company Location (2006-2017)', fontsize=20)
plt.xlabel('Company Location', fontsize=15)
plt.ylabel('Rating (1-5)', fontsize=15)
plt.margins(.03)

plt.show()
```

From the scatterplot above, **Chile** has the highest rating of chocolate while **India** has the lowest **on average**. However, as shown below, the company located in **Chile** and **India** has only **2** and **1** observations, respectively. It would be hard to conclude that they have the best or worst chocolate bars. For example, **Belgium** is famous for their chocolate, but they did have chocolate bar scoring **1** in this dataset.
