Importing the data set:

```python
df = pd.read_csv("/kaggle/input/the-bread-basket/bread basket.csv") 
df.head()
```

Creating a sub list of items having same transaction id in a list

```python
# https://stackoverflow.com/questions/65154530/how-to-combine-values-having-same-value-in-some-columns
from collections import defaultdict
refHash = defaultdict(list)

for i,id in enumerate(df['Transaction'].values):
    refHash[id] += [df.loc[i,'Item']]

lst = list(refHash.values())
```

Training the APRIORI machine

```python
from apyori import apriori
rules = apriori(transactions= lst, min_support = 0.002, min_lift = 3)
results = list(rules)
results
```

Better visualization of the output from the model(RULES)

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
