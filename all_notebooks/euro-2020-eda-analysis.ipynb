## team general statistics.

```python
countrys =df.Country.value_counts().index
# ['England', 'Italy', 'Spain', 'France', 'Germany', 'Portugal', 'Belgium','Denmark', 'Switzerland', 'Wales', 'Netherlands', 'Ukraine', 'Austria','Croatia', 'Czech Republic', 'SwitzerlandF', 'Sweden', 'Poland']

values=df.Country.value_counts().values
#[7, 7, 5, 5, 5, 5, 4, 4, 2, 2, 2, 2, 2, 2, 2, 1, 1, 1]
  
# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(values, labels = countrys)
  
# show plot
plt.show()
```

* By looking at these numbers, it can be seen that the best players are in the best teams.

```python
# Number of players by position played.

plt.figure(figsize=(15,10))
sns.barplot(x=df['Position'].value_counts().index, y=df['Position'].value_counts().values)
plt.xticks(rotation= 45)
plt.xlabel('position')
plt.ylabel('number of player')
plt.title('football player and position chart .')
```

```python
# Number of goals scored by country.

country_goals=[]
for a in countrys:
    country_goals.append(sum(df.Goals[df.Country==a]))

plt.figure(figsize=(15,10))
sns.barplot(x=countrys, y=country_goals)
plt.xticks(rotation= 45)
plt.xlabel('countrys')
plt.ylabel('goals')
plt.title('total number of goals by countries')
```

```python
## Let's see how goals are scored.

way_of_scoring =df.iloc[:,4:8]

x = countrys

trace1 = {
  'x': x,
  'y': way_of_scoring['Right foot goals'],
  'name': 'Right foot goals',
  'type': 'bar'
};
trace2 = {
  'x': x,
  'y': way_of_scoring['Left foot goals'],
  'name': 'Left foot goals',
  'type': 'bar'
};
trace3 = {
  'x': x,
  'y': way_of_scoring['Header goals'],
  'name': 'Header goals',
  'type': 'bar'
};
data = [trace1, trace2,trace3];
layout = {
  'xaxis': {'title': 'countries'},
  'barmode': 'relative',
  'title': 'goal scoring chart'
};
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

```python
attempts_scoring=df

#total attempts for country
country_Total_attempts=[]
for a in countrys:
    country_Total_attempts.append(sum(attempts_scoring['Total attempts'][df.Country==a]))
    
#on target for country
country_On_target=[]
for a in countrys:
    country_On_target.append(sum(attempts_scoring['On target'][df.Country==a]))

#off target for country
country_Off_target=[]
for a in countrys:
    country_Off_target.append(sum(attempts_scoring['Off target'][df.Country==a]))
```

```python
f,ax = plt.subplots(figsize = (9,15))

sns.barplot(x=country_Total_attempts,y=countrys,color='green',alpha = 0.5,label='Total attempts' )
sns.barplot(x=country_On_target,y=countrys,color='red',alpha = 0.5,label='On target' )
sns.barplot(x=country_Off_target,y=countrys,color='blue',alpha = 0.5,label='Off target' )

ax.legend(loc='lower right',frameon = True)     
ax.set(xlabel='attempts', ylabel='countries',title = "attempts statistics ")
```

```python
#total yellow card
country_yellow_card=[]
for a in countrys:
    country_yellow_card.append(sum(df['Yellow cards'][df.Country==a]))

#total red card
country_red_card=[]
for a in countrys:
    country_red_card.append(sum(df['Red cards'][df.Country==a]))
    
#total Fouls committed
country_Fouls_committed=[]
for a in countrys:
    country_Fouls_committed.append(sum(df['Fouls committed'][df.Country==a]))
```

```python
tr1 = go.Bar(
                x = countrys,
                y = country_Fouls_committed,
                name = "Fouls committed",
                marker = dict(color = 'rgba(100, 250, 128, 0.5)',
                              line=dict(color='rgb(0,0,0)',width=1.5)),
                text = countrys)

tr2 = go.Bar(
                x = countrys,
                y =country_yellow_card,
                name = "yellow card",
                marker = dict(color = 'rgba(255, 255, 128, 0.5)',
                             line=dict(color='rgb(0,0,0)',width=1.5)),
                text = countrys)

tr3 = go.Bar(
                x = countrys,
                y = country_red_card,
                name = "red card",
                marker = dict(color = 'rgba(255, 100, 128, 0.5)',
                              line=dict(color='rgb(0,0,0)',width=1.5)),
                text = countrys)


data = [tr1, tr2 , tr3]
layout = go.Layout(barmode = "group")
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```
