```python
import pandas as pd
df = pd.read_csv('../input/advertising-data/Advertising.csv')
df =df.iloc[:,1:len(df)]
df.head()
```

```python
df.info()
```

```python
import seaborn as sns
sns.jointplot(x = 'TV', y = 'Sales', data = df, kind = 'reg');
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[['TV']]
X.head()
```

```python

y = df[['Sales']]
y.head()
```

```python
reg = LinearRegression()
model = reg.fit(X,y)
```

```python
dir(model)
```

intercept ifadesi denklemin sabitidir, coef ifadesi katsayidir.

```python
model.intercept_
```

```python
model.coef_
```

```python
#rkare : bagimli degiskendeki degisikligin, bagimsiz degiskenlerce aciklanma yuzdesidir. 
#Bu ornekte satislardaki degisikligin yaklasik %60 i bagimsiz degiskence aciklanmaktadir.
model.score(X,y)
```

### Artiklar ve Makine Ogrenmesindeki Onemi
#### MSE: Hata Kareler Ortalamasi
#### Hata Kareler Ortalamasi Karekoku

```python
gercek_y = y[0:10]
```

```python
tahmin_y = pd.DataFrame(model.predict(X)[0:10])
```

```python
hatalar = pd.concat([gercek_y,tahmin_y],axis = 1)
hatalar.columns = ['gercek_y','tahmin_y']
hatalar
```

```python
hatalar['hata'] = hatalar['gercek_y'] - hatalar['tahmin_y']
hatalar
```

```python
hatalar['hata_kareler'] = hatalar['hata']**2
hatalar
```

```python
import numpy as np
np.mean(hatalar['hata_kareler'])#10 gozlem icin hata kareler ortalamasi el yordamiyla bulunmus oldu
```
