```python
import numpy as np
import pandas as pd
import math
```

```python
df=pd.read_csv("../input/flight-take-off-data-jfk-airport/M1_final.csv")
print(df)
```

```python
print(df.columns)
```

```python
df=df.drop("TAIL_NUM",axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
le.fit(df.iloc[:,3])
df.iloc[:,3]=le.transform(df.iloc[:,3])
le.fit(df.iloc[:,4])
df.iloc[:,4]=le.transform(df.iloc[:,4])
le.fit(df.iloc[:,14])
df.iloc[:,14]=le.transform(df.iloc[:,14])
le.fit(df.iloc[:,18])
df.iloc[:,18]=le.transform(df.iloc[:,18])
```

```python
Y=df.iloc[:,21]
print(np.shape(Y))
```

```python
X=df.drop(["TAXI_OUT"],axis=1)
print(np.shape(X))
```

```python
from sklearn.model_selection import train_test_split
X_train,x_test,Y_train,y_test=train_test_split(X,Y,test_size=.1)
print(np.shape(X_train),np.shape(x_test),np.shape(Y_train),np.shape(y_test))
```

```python
x_test.index=np.arange(len(x_test))
X_train.index=np.arange(len(X_train))
y_test.index=np.arange(len(y_test))
Y_train.index=np.arange(len(Y_train))
```

Linear Regression

```python
from sklearn import linear_model
LM=linear_model.LinearRegression()
LM.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
rmse1=pow(mean_squared_error(LM.predict(x_test),y_test),.5)
print(rmse1)
```

Lasso Regression

```python
from sklearn.linear_model import Lasso
LR1=Lasso(alpha=1,max_iter=1000,tol=.01)
LR1.fit(X_train,Y_train)
```

```python
rmse2=pow(mean_squared_error(LR1.predict(x_test),y_test),.5)
print(rmse2)
```

Ridge Regression

```python
from sklearn.linear_model import Ridge
RR1=Ridge(alpha=1,max_iter=1000,tol=.01)
RR1.fit(X_train,Y_train)
```

```python
rmse3=pow(mean_squared_error(RR1.predict(x_test),y_test),.5)
print(rmse3)
```

KNN Algorithm

```python
from sklearn.neighbors import KNeighborsRegressor
knn1=KNeighborsRegressor(n_neighbors=500)
knn1.fit(X_train,Y_train)
```

```python
rmse4=pow(mean_squared_error(knn1.predict(x_test),y_test),.5)
print(rmse4)
```
