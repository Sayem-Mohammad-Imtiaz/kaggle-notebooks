There is lot of visualization techniques for visualization this dataset. By Visualizations we can see graph of Date v/s ItemDesciption. Which Item sold on Which date ??

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

There you can see a lot of data here. Big that can not be readable from this graph. Let's visualize something that is readable :::::

```python
plt.plot(data.Date, data.itemDescription)
plt.title('Visualization - 1')
plt.xlabel("Item Description")
plt.ylabel("Date")
plt.show()
```

Oh, its great. This visualization has hover data so that you can see the visualization as well as the data under it. Let's try another one :::

```python
fig = px.scatter(data, x="Date", y="itemDescription", 
                 color="Member_number", 
                 hover_data=['Member_number'], 
                 title = "Visualizations - 2")
fig.show()
```

This visualization tells us date vs numbers increments and decrements you can see.

```python
import plotly.graph_objects as go
fig2 = go.Figure(data=go.Scatter(
    y = data['Date'],
    mode='markers',
    marker=dict(
        size=16,
        color=data['Member_number'], #set color equal to a variable
        colorscale='Viridis', # one of plotly colorscales
        showscale=True
    )
))
fig2.update_layout(title='Visualization - 2',
                  yaxis_title="Date")
fig2.show()
```

```python
fig3 = px.scatter(data, x="Date", y="itemDescription", 
                 color="Member_number",
                 size='Member_number', 
                 hover_data=['Member_number'], 
                 title = "Visualization - 4")
fig3.show()
```

```python
fig2 = go.Figure(data=go.Scatter(x=data['Date'],
                                y=data['itemDescription'],
                                mode='markers',
                                marker_color=data['Member_number'],
                                text=data['Member_number'])) # hover text goes here

fig2.update_layout(title='Visualization - 5')
fig2.show()
```
