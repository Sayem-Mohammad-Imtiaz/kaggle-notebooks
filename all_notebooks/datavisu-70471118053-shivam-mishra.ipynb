 ***Indexing***

Command used to index the data i.e. we extract a particular data from the dataset.
In the below example we indexed the data for 4th row and so it shows us the 4th row data.

```python
df.iloc[4]
```

This can also be indexed via names(provided in the dataset) both for colomns and rows.
In the below example we indexed the data on the basis of credit limit....So the output shows us all the rows for credit limit.

```python
df.loc[:,'Credit_Limit']
```

Indexing can also be done for multiple rows and colomns.
here for example we indexed the data for 3 rows i.e. 1, 2, 3.

```python
df.iloc[[1,2,3]]
```

Similar example for indexing on the basis of names (can be both colomns and rows).
Here we index on the basis of Income category and credit limit.

```python
df.loc[:,['Income_Category', 'Credit_Limit']]
```

***SLICING***

 Slicing means taking elements from one given index to another given index.
 So for example we wants to see the results from index 5(row) inthe dataset.

```python
df[5:]
```

Slicing can also done for both rows and colomns and also it can be done by the names provided in the data.
So for example we want 5 rows on the basis of Income category and credit limit we have the following code.

```python
df.loc[:,['Income_Category', 'Credit_Limit']][:5]
```

***SUBSET OF THE DATA***

Suppose we want a specific part of data...Such as people having income less than $40k, we use the following peice of code.


```python
df[df.Income_Category == "Less than $40K"]
```

```python
df['Customer_Age'].value_counts().sort_index().plot.line()
```

```python
df['Customer_Age'].value_counts().sort_index().plot.area()
```

4. HISTOGRAM

```python
df['Credit_Limit'].plot.hist()
```
