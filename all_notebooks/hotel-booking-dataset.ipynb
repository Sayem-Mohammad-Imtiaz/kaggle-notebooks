```python
## Q6 - Does it have any missing data? If the answer is positive, which column have the most missing data?
# Solution source: https://bit.ly/2TyM0Dq

# Q4-1 - Checking missing data
checkMD = df.isnull().values.any();

# Q4-2 - The column with most missing data
columnMD = max(df.isnull());

# Get the output
pd.DataFrame({
    'check_missing_data_existance': checkMD,
    'column_with_maximum_missing_data': columnMD.upper()
},index=['0'])
```

```python
# Q7 - Drop 'company' column from DataFrame.
df.drop('company',axis=1).columns.tolist()
```

```python
# Q8 - Which country have the most passengers? Define the top 10 countries.
df['total_passengers'] = (df['adults'] + df['children'] + df['babies']) - df['is_canceled'];

df[['country','adults','children','babies','is_canceled','total_passengers']].groupby('country').sum().nlargest(10,'total_passengers').reset_index()
```

```python
# Q9 - Define the name of the passenger who have the maximum average daily rate (ADR). How much is its price?
name = df.name[df['adr'].idxmax()];
price = df['adr'].max();

dict = {
    'name': name,
    'price': price
};

pd.DataFrame(dict,index=['Max(ADR)'])
```

```python
# Q9 - Average of total ADRs with 2 decimals.
df['adr'].mean().round(3)
```

```python
#Q 10find the statistical finding
df.describe()
```

```python
# Q11 - Define the name and e-mail of people who had 5 special requests.
df[df['total_of_special_requests'] == 5][['name','email','credit_card','phone-number']]
```

```python
# Q12 - Define the average of number of nights stayed.
df['total_stays_in_nights'] = df['stays_in_week_nights'] + df['stays_in_weekend_nights'];

df['total_stays_in_nights'].mean().round(2)
```

```python
# Q13 - Which first names have the most frequency of last name? Define 5 most frequent family.
df['name'].apply(lambda lname:lname.split()[1]).value_counts().head().to_frame('value_counts')
```

```python
# Q14 - Define the people whom reserved a hotel with most number of babies and children.
df['total_babies_and_children'] = df['babies'] + df['children'];

df[['name','email','phone-number','total_babies_and_children']].iloc[df['total_babies_and_children'].idxmax()].to_frame('content')
```

```python
# Q15 - Define the phone number of regions which had the most reservations.
df['phone-number'].apply(lambda phone_code:phone_code[:3]).value_counts().nlargest(3).to_frame('value_counts')
```
