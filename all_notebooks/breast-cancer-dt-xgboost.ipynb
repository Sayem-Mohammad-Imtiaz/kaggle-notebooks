# ****Data Loading and Visualisation*****

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
data = pd.read_csv('../input/breast-cancer-wisconsin-data/data.csv')
data.head()
```

```python
sb.set_theme(style="darkgrid")
sb.countplot(data['diagnosis'])

```

```python
sb.heatmap(data.corr())
```

# ***Data Preprocessing***

```python
data = data.drop(['id', 'Unnamed: 32'], axis=1)
diagnosis = data['diagnosis']
lb = preprocessing.LabelEncoder()
lb.fit(diagnosis)
diagnosis = lb.transform(diagnosis)
print(diagnosis)


data = data.drop(['diagnosis'], axis=1)

```

# ****XGBOOST_Model****

```python
from sklearn import metrics
from sklearn.model_selection import train_test_split
from lightgbm import LGBMClassifier
import lightgbm as lgb
from sklearn.ensemble import GradientBoostingClassifier
```

```python
x_train= data; x_test = diagnosis
x_train, x_test, y_train, y_test = train_test_split(x_train, x_test, test_size=0.30)

 
clf = GradientBoostingClassifier(learning_rate=1.0).fit(x_train, y_train)
clf.score(x_test, y_test)
```

```python
expected_y  = y_test
predicted_y = clf.predict(x_test)

print(metrics.classification_report(expected_y, predicted_y))
```

```python
print(metrics.confusion_matrix(expected_y, predicted_y))
```

```python
print(metrics.r2_score(expected_y, predicted_y))
print(metrics.mean_squared_log_error(expected_y, predicted_y))
```

# **DECISION_TREE Model**

```python
from sklearn import tree
```

```python
dt = tree.DecisionTreeClassifier()
dt = dt.fit(x_train, y_train)
dt.score(x_test, y_test)
```

```python
expected_y  = y_test
predicted_y = dt.predict(x_test)

print(metrics.classification_report(expected_y, predicted_y))
```

```python
print(metrics.r2_score(expected_y, predicted_y))
print(metrics.mean_squared_log_error(expected_y, predicted_y))
```
