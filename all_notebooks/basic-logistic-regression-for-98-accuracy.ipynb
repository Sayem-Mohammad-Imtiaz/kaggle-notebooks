```python
# Lets first do feature scaling on all the numerical columns
```

```python
data_X = data.drop('diagnosis', axis =1)
data_y = data['diagnosis']
```

```python
data_X.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled = scaler.fit_transform(data_X)
```

```python
df_x = scaled.copy()
```

```python
df_x
```

```python
# Coverting it back to dataframe format
df_x = pd.DataFrame(scaled, columns =['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean',
       'smoothness_mean', 'compactness_mean', 'concavity_mean',
       'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean',
       'radius_se', 'texture_se', 'perimeter_se', 'area_se', 'smoothness_se',
       'compactness_se', 'concavity_se', 'concave points_se', 'symmetry_se',
       'fractal_dimension_se', 'radius_worst', 'texture_worst',
       'perimeter_worst', 'area_worst', 'smoothness_worst',
       'compactness_worst', 'concavity_worst', 'concave points_worst',
       'symmetry_worst', 'fractal_dimension_worst'])
```

```python
df_x
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
data_Z = [1 if each == "M" else 0 for each in data_y] # encoding the class values
```

```python
data_Z = pd.DataFrame(data_Z, columns = ['Diag'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_x, data_Z, test_size=0.3)
```

```python
clf = LogisticRegression(C = 0.40, max_iter = 100)
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_train,y_train) # Score on training data
```

```python
clf.score(X_test,y_test) # Score on testing data, lets do cross -validation to check the average accuracy
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(clf,df_x, data_Z, cv =30).mean()
```
