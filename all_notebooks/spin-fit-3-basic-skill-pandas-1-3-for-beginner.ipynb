```python
# From dict of lists, dict of arrays, dict of series

names = ['Kym', 'sena', 'paul', 'smith']
ages = ['32', '30', '22', '58']
height = ['183', '165', '187', '175']
```

```python
dict1 = {
    'name': names,
    'age': ages,
    'height': height
}
df1 = pd.DataFrame(dict1)
df1
```

```python
dict2 = {
    'name': np.array(names),
    'age': np.array(ages),
    'height': np.array(height)
}
df2 = pd.DataFrame(dict2)
df2
```

```python
dict3 = {
    'name': pd.Series(names),
    'age': pd.Series(ages),
    'height': pd.Series(height)
}
df3 = pd.DataFrame(dict3)
df3
```
