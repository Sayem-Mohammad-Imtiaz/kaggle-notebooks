### Lets's check the Outliers:

```python

def outliers(df):
    out=[]
    q1=df.quantile(0.25)
    q3=df.quantile(0.75)
    iqr=q3-q1
    upper_value=q3+1.5*iqr
    lower_value=q1-1.5*iqr
    for x in df:
        if x > upper_value or x< lower_value:
            out.append(x)
    print("OUTLIERS ARE : \n",out)
    
    
    
    
```

```python
editions=df["Name"].value_counts().sort_values(ascending=False)[0:10].reset_index()
editions.columns=["Name","editions"]
plt.figure(figsize=(10,6))
sns.barplot(x=editions["Name"],y=editions["editions"])
plt.xticks(rotation=90)
plt.title("TOP TEN BOOKS WITH MAX. NUMBER OF EDITIONS",fontsize=20)
plt.show()
```

```python
Authors_with_max_books=df["Author"].value_counts().sort_values()[-10:].reset_index()
Authors_with_max_books.columns=["Author","Number_of_books"]
plt.figure(figsize=(10,6))
ax=sns.barplot(x=Authors_with_max_books["Author"],y=Authors_with_max_books["Number_of_books"])
plt.xticks(rotation=90)
plt.title("TOP TEN AUTHORS WITH MAX NUMBER OF BOOKS",fontsize=20)
plt.show()


```

## 02. Bivariate Analysis:

```python
top_reviews=df.groupby(df["Author"])["Reviews"].max().sort_values()[-10:].reset_index()
plt.figure(figsize=(10,6))
ax=sns.barplot(x=top_reviews["Author"],y=top_reviews["Reviews"])
plt.xticks(rotation=90)
plt.title("TOP TEN AUTHORS WITH HIGHEST NUBER OF REVIEWS",fontsize=20)

for i in ax.patches:
    ax.text(i.get_x()+.25,i.get_height()+2.3,str(int((i.get_height()))))
            #rotation=0,fontsize=15,color='black')
```

```python
print("TOP TEN BOOKS WITH HIGHEST RATING")
df.groupby(df["Name"])["User Rating"].max().sort_values()[-10:]
```

```python
books_with_high_price=df.groupby(df["Name"])["Price"].max().sort_values()[-10:].reset_index()
plt.figure(figsize=(10,6))
ax=sns.barplot(x=books_with_high_price["Name"],y=books_with_high_price["Price"])
plt.xticks(rotation=90)
plt.title("TOP TEN BOOKS WITH HIGHEST PRICE",fontsize=20)
for i in ax.patches:
    ax.text(i.get_x()+.25,i.get_height()+2.3,str(int((i.get_height()))),
            rotation=0,fontsize=15,color='black')
```

## 03.Multivariate Analysis:

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
Gen=df[["Genre","User Rating"]].value_counts().reset_index()
Gen.columns=["Genre","User Rating","Count"]
plt.figure(figsize=(15,7))
sns.barplot(hue=Gen["Genre"],x=Gen["User Rating"],y=Gen["Count"])
plt.title("BASED ON GENRE COUNT OF RATING",fontsize=20)
plt.show()
```

```python
top=df[["Name","Price","Author"]].sort_values(by="Price")[-20:].reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x=top["Name"],y=top["Price"],hue=top["Author"])
plt.xticks(rotation=90)
plt.title("BOOKS WITH HIGH PRICE AND THEIR AUTHORS ",fontsize=20)
plt.show()
```

```python
top=df[["Name","Reviews","Author"]].sort_values(by="Reviews")[-20:].reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x=top["Name"],y=top["Reviews"],hue=top["Author"])
plt.xticks(rotation=90)
plt.title("BOOKS WITH HIGHEST REVIEWS AND THEIR AUTHORS ",fontsize=20)
plt.show()
```

```python
fiction=df[df["Genre"]=="Fiction"]
top=fiction[["Author","Name","Reviews"]].sort_values("Reviews")[-20:].reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x=top["Name"],y=top["Reviews"],hue=top["Author"])
plt.title("TOP FICTION BOOKS WITH AUTHOR AND THEIR REVIEWS",fontsize=20)
plt.xticks(rotation=90)
plt.show()
```

```python
fiction=df[df["Genre"]=="Non Fiction"]
top=fiction[["Author","Name","Reviews"]].sort_values("Reviews")[-20:].reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x=top["Name"],y=top["Reviews"],hue=top["Author"])
plt.title("TOP NON FICTION BOOKS WITH AUTHOR AND THEIR REVIEWS",fontsize=20)
plt.xticks(rotation=90)
plt.show()
```
