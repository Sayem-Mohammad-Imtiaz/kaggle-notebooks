# 1) Which US state has the highest number of accidents

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../input/us-accidents/US_Accidents_Dec20.csv")
df.head()
```

**Check if there are null values**


```python
df.isnull().sum()
```

**There is no null values in states**

**We can group by states and see number of accidents**

```python
states = df.groupby(['State']).count().reset_index()
```

```python
states.head()
```

**We can find the state with the maximum number of accidents**

**We can also visualise results for better understanding (comparison)**

```python
states.loc[states['ID'] == states['ID'].max()]
```

```python
#states.plot.bar(x = 'State', y = 'ID', figsize = (16,8))

f, ax = plt.subplots(figsize=(18, 10))
sns.barplot(x = 'State', y = 'ID', data = states)
plt.xlabel("States", labelpad = 10, fontsize=12,weight='bold')
plt.ylabel("Number of Accidents", labelpad = 10,fontsize=12, weight='bold')
plt.title("Accidents in States", fontsize=15, weight='bold')
```

# 2) Description of the accidents that usually occur in that state. (CA)

Create new table only for CA

```python
CA = df.loc[df['State'] == 'CA']
CA
```

**We need to set up number of accidents for each type**

For this, we need to iterate through the list of 'accident types' and find the number of each accident type.
Then, we create DataFrame and plot it

```python

accident_types = ['Amenity', 'Bump', 'Crossing','Give_Way', 'Junction', 'No_Exit', 'Railway', 'Roundabout',
                  'Station', 'Stop', 'Traffic_Calming', 'Traffic_Signal', 'Turning_Loop']
lst = []


for i in range(len(accident_types)):
    num = CA.loc[CA[f'{accident_types[i]}']].count() # num is the number of particular accident type 
    # ID represents the correct number of accidents of each type, as other parameters could be null/na
    #print(num)

    lst.append(num[1]) # we need only first value corresponding to ID for each accident type
    
lst
        
```

**Create a Dataframe from dictionary and plot it**

```python
data = {'Accident types': accident_types, 'Number of Accidents': lst}
accidents = pd.DataFrame.from_dict(data)
accidents = accidents.sort_values('Number of Accidents', ascending = False).reset_index()
accidents
```

```python
f, ax = plt.subplots(figsize=(18, 10))
sns.barplot(y='Number of Accidents', x = 'Accident types', data=accidents)
plt.xlabel("Accident types", labelpad = 10, fontsize=12,weight='bold')
plt.ylabel("Number of Accidents", labelpad = 10,fontsize=12, weight='bold')
plt.title("Accidents in California", fontsize=15, weight='bold')
```
