```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mp
```

```python
file = '../input/highest-grossing-movies-of-actors/Actor data.csv'
```

```python
df = pd.read_csv(file)
```

```python
df.head(26)
```

```python
a = df['Actor'].value_counts()
print(a)
```

```python
%matplotlib inline
plt.rcParams['figure.figsize']=30,20
```

```python
a f= df.groupby('Actor') ['Film'].sum()
af = dict(a)
af
```

```python
a.plot(kind = 'bar',color = 'greenyellow')
plt.xticks(fontsize = 25,rotation = 65)
plt.yticks(fontsize = 20)
plt.xlabel("Actor name",fontsize = 30)
plt.ylabel("Number of movies",fontsize = 30)
plt.grid(color='#95a5a6', linestyle='--', linewidth=2, axis='y', alpha=0.7)
plt.title("Highest-earning actors for a single production",fontsize = 50)
```

```python
b = df['Year'].value_counts()
print(b)
b.plot( c ='Black' , ls='', marker = 'o', ms=30)
plt.show
plt.xticks(fontsize = 30)
plt.yticks(fontsize = 30)
plt.xlabel("Year",fontsize = 35)
plt.ylabel("Number of movies",fontsize = 35)
plt.title("Year and number of best selling movies",fontsize = 50)
```

```python
c = df['Total income'].value_counts()
print(c)
c.plot(kind = 'bar',color = 'palevioletred')
plt.xticks(fontsize = 20,rotation = 70)
plt.yticks(fontsize = 20)
plt.xlabel("Total Income",fontsize = 25)
plt.ylabel("Number of movies",fontsize = 25)
plt.title("Total income and number of best selling movies",fontsize = 50)
```

```python
d = df['Salary'].value_counts()
print(d)
plt.plot(d, c ='Blue' , ls='', marker = 'o', ms=30)
plt.xticks(fontsize = 25)
plt.yticks(fontsize = 25)
plt.xlabel("Salary ",fontsize = 30)
plt.ylabel("Number of actors",fontsize = 30)
plt.title("Salary and the number of actors",fontsize = 50)
```
