# Preprocessing

```python
import re
```

```python
tweets=" ".join(tweet for tweet in df['tweet']) #string containing all the tweets
tweets[0:1000]
```

```python
#removing all the mentions and users
tweets=re.sub(r'@\w+','',tweets)
tweets[0:1000]
```

```python
#removing all the urls and replacing it with _URL_
tweets=re.sub(r'http\S+','',tweets)
tweets[0:1000]
```

```python
#removing all the #
tweets=re.sub(r'#','',tweets)
tweets[0:1000]
```

```python
#converting into lower
tweets=tweets.lower()
tweets[0:1000]
```

```python
#removing all the puntuations
tweets = re.sub(r'[^\w\s]','',tweets)
tweets[0:1000]
```

```python
corpus=[]
tweets=''.join(tweets)
corpus.append(tweets)
```

# Visualizations 

```python
df['target'].value_counts().sort_values(ascending=False).head(10).plot(kind='bar',figsize=(12,6))
plt.xlabel('Target')
plt.ylabel('Count of tweets')
plt.title('Top 10 targets of Trump')
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
wc = WordCloud(background_color='black',max_words=100)
fig=plt.figure(figsize=(10,10))
wc.generate(str(corpus))
plt.imshow(wc, interpolation="bilinear")
plt.title('Top words which trump used')
plt.axis('off')
plt.show()
```
