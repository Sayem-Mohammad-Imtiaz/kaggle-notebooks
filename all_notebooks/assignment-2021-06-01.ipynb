# Create a global map for COVID Vaccinations (The gap between the actual number of vaccines & how much vaccines are used).

I visualized two maps.
①Total number of people fully vaccinated per hundred - percent between population fully immunized and total population 
②Total vaccinations per hundred - percent between vaccination number and total population

I create these two at the same time to compare "how many vaccines are imported" with "how many of them are actually utilized". (So I set the same color & range to compare.)

What I found interesting about this comparison is this; although Canada has already imported considerable amount of vaccines (63.22% of the population), only 5.5% is fully vaccinated. 

```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
print(df.columns)
# can load a data frame and construct a map
#range_color=(0.0,40.0) means the min=0 and the max=40
fig1 = px.choropleth(df,locations='country',locationmode='country names',color='people_fully_vaccinated_per_hundred',hover_name='country',animation_frame='date',color_continuous_scale='rainbow',range_color=(0.0,100.0))
#00:50:49~
fig1.update_layout(title_text="Percentage of the country's total population that are fully vaccinated",title_x=0.43)
fig1.show()#00:47:10~

plotly.offline.plot(fig1,filename='vaccinations_completed.html')

fig2 = px.choropleth(df,locations='country',locationmode='country names',color='total_vaccinations_per_hundred',hover_name='country',animation_frame='date',color_continuous_scale='rainbow',range_color=(0.0,100.0))
fig2.update_layout(title_text="Percentage of the country's vaccinations",title_x=0.43)
fig2.show()

plotly.offline.plot(fig2,filename='vaccinations_number.html')
```

# Create a global map of earthquake occurrences worldwide.

Data source is here↓

https://earthquake.usgs.gov/earthquakes/search/

```python
df = pd.read_csv('../input/202106-recent-earthquake/query.csv',header=0)
df.index = pd.to_datetime(df['time'])

#eliminate the minute and second
df['time'] = df.index.strftime('%Y-%m-%d %H:00:00')
df = df.sort_index()
fig = px.scatter_geo(df,lat='latitude', lon='longitude',color='mag',animation_frame='time',color_continuous_scale='rainbow',range_color=(1.0,3.0))
fig.show()
```
