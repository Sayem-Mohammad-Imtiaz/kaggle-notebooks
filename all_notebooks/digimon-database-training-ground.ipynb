<h1>Digimon Database Training Ground </h1>
This dataset will help you theorycraft the ultimate team as well as ask interesting questions.

– Which set of moves will get the best ratio of attack power to SP spent?

– Which team of 3 digimon have the highest attack? Defense?

– What’s the tradeoff between HP and SP?

– Are some types over- or under-represented?

– Both the moves and support skills have short text descriptions. Can an NLP analysis reveal underlying clusters of moves?

– Are different types and attributes evenly represented across stages?


```python
BASE_PATH  = "../input/digidb"
os.listdir(BASE_PATH)
DigiDB_movepath = os.path.join(BASE_PATH, os.listdir(BASE_PATH)[0])
DigiDB_supportpath = os.path.join(BASE_PATH, os.listdir(BASE_PATH)[1])
DigiDB_digimonpath = os.path.join(BASE_PATH, os.listdir(BASE_PATH)[2])
DigiDB_move = pd.read_csv(DigiDB_movepath)
DigiDB_support = pd.read_csv(DigiDB_supportpath)
DigiDB_digimon = pd.read_csv(DigiDB_digimonpath)
```

```python
import matplotlib.pyplot as plt
#attributes_volume = DigiDB_digimon.groupby("Attribute",  as_index = False).aggregate({'Number':'count'})
#attributes_volume = attributes_volume.sort_values("Number")
#type_volume = DigiDB_digimon.groupby("Type",  as_index = False).aggregate({'Number':'count'})
#type_volume = type_volume.sort_values("Number")
##################
fig, axs = plt.subplots(9, 2, figsize=(22, 35), sharey=False, constrained_layout=True) # Инициализируем полотно для рисования графиков
u = 0

for i in  DigiDB_digimon['Stage'].unique():
    attributes_volume = DigiDB_digimon[DigiDB_digimon['Stage'] == i].groupby("Attribute",  as_index = False).aggregate({'Number':'count'})
    attributes_volume = attributes_volume.sort_values("Number").fillna(0)
    type_volume = DigiDB_digimon[DigiDB_digimon['Stage'] == i].groupby("Type",  as_index = False).aggregate({'Number':'count'})
    type_volume = type_volume.sort_values("Number").fillna(0)
    rgb = np.random.rand(3,)
    axs[u, 0].bar(attributes_volume.Attribute,attributes_volume.Number, color = [rgb])
    axs[u, 1].bar(type_volume.Type, type_volume.Number, color = [rgb])
    axs[u, 0].set_title(i + ' Number of digimons for each attribute')
    axs[u, 1].set_title(i + ' Number of digimons for each type')
    u += 1
    
attributes_volume = DigiDB_digimon.groupby("Attribute",  as_index = False).aggregate({'Number':'count'})
attributes_volume = attributes_volume.sort_values("Number")
type_volume = DigiDB_digimon.groupby("Type",  as_index = False).aggregate({'Number':'count'})
type_volume = type_volume.sort_values("Number")
axs[8, 0].bar(attributes_volume.Attribute,attributes_volume.Number)
axs[8, 1].bar(type_volume.Type,type_volume.Number)
axs[u, 0].set_title(' Number of digimons for each attribute')
axs[u, 1].set_title(' Number of digimons for each type')

#axs[2].bar(attributes_volume.Attribute,attributes_volume.Number)
fig.suptitle('Categories of Digimons', size = 20)
```

```python
import matplotlib
import matplotlib.patches as mpatches
DigiDB_digimon.head()
DigiDB_digimon["Type"] = DigiDB_digimon["Type"].astype('category')
DigiDB_digimon["Type_cat"] = DigiDB_digimon["Type"].cat.codes
DigiDB_digimon.dtypes
fig1, ax1 = plt.subplots(figsize=(8,8))
colors = ['red','green','blue','purple']
Digimon_types = list(dict(enumerate(DigiDB_digimon["Type"].cat.categories)).values())
ax1.scatter(DigiDB_digimon['Lv50 Atk'], DigiDB_digimon['Lv50 SP'], c=DigiDB_digimon['Type_cat'], cmap=matplotlib.colors.ListedColormap(colors))
patches = list()
for index, color1 in enumerate(colors): 
    patches.append(mpatches.Patch(color=color1, label=Digimon_types[index]))
    ax1.legend(handles=patches)

fig1.suptitle('The tradeoff between HP and SP', size = 25)
ax1.set_xlabel('Attack', fontsize=20, labelpad=20)
ax1.set_ylabel('SP',  rotation=0, fontsize=20, labelpad=20)
```
