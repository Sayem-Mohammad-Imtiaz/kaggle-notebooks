```python
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

Load and Observe Data

```python
df = pd.read_csv("/kaggle/input/salary-data-simple-linear-regression/Salary_Data.csv")
df.head()
```

```python
df.describe()
```

```python
df.plot()
```

```python
sns.scatterplot(data = df, x='YearsExperience', y='Salary')
```

```python
X = df['YearsExperience']
y = df['Salary']

print(X)
print(y)
```

Split the Data into Test and Train data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train = X_train.values.reshape(-1,1)
X_test = X_test.values.reshape(-1,1)
y_train = y_train.values.reshape(-1,1)
```

Creating Model and Training it

```python
linear_model = LinearRegression()
```

```python
linear_model.fit(X_train, y_train)
```

Make Predictions

```python
predictions = linear_model.predict(X_test)
```

Calculating Error

```python
error = mean_squared_error(y_test, predictions)
```

```python
print('Intercept of the model:',linear_model.intercept_)
print('Coefficient of the line:',linear_model.coef_)
```
