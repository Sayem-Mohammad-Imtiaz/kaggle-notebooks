```python
def per_country(country_name):
    f, axes = plt.subplots(1, 2,figsize=(10,6))
    country_data = data[data["location"]==country_name]
    sns.scatterplot(x="total_deaths",y="total_cases",data=country_data ,ax=axes[0])
    sns.scatterplot(x="population",y="total_deaths",data=country_data,ax=axes[1])
#     plt.tight_layout()
    country_data.head(12)
    
per_country("Afghanistan")
```
