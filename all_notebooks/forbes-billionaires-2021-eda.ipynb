**Exploratory data analysis on billionaires data by Forbes. Feito por Kayo Ricardo**

Which countries have the highest number of billionaires

```python
top10_countries = df['Country'].value_counts().head(10)
top10_countries
```

```python
top10_countries.plot.barh(title='Countries with more billionaires').invert_yaxis()
```

```python
df.groupby('Country')['NetWorth'].sum().sort_values(ascending=False).head(10).plot.barh(title = 'Wealthness by country',xlabel='').invert_yaxis()
```

Which colleges in Brazil the brazillians billionaires studied

```python
br['Education'].value_counts()
```

How many studied at Universidade de São Paulo

```python
# I used only "Sao Paulo" because "Universidade" can appear as "University" or as the name of the Institute
usp = br[br['Education'].str.contains("Sao Paulo") == True]
usp
```

```python
# Looks like Universidade Federal do Rio de Janeiro is very frequent

ufrj = br[br['Education'].str.contains("Rio de Janeiro") == True]
ufrj
```

Let's see the most frequent sources

```python
banking = br[br['Source'].str.contains("banking") == True]
banking
```

```python
beer = br[br['Source'].str.contains("beer") == True]
beer
```

```python
retail = br[br['Source'].str.contains("retail") == True]
retail
```

```python
hospital = br[(br['Source'].str.contains("hospital") == True) | (br['Source'].str.contains("health"))]
hospital
```

Conclusions:

The average age of brazillian billionaires is very similar to the world billionaires average age (62 x 63). However, the average of the world top 10 billionaires drops to 60 years old while in Brazil the average go up to 67 years old.

The countries with the highest number of billionaires are United States and China. Brazil, a developing country, appears in front of Canada and United Kingdom, developed countries. But of course the total population of Brazil is way bigger than the population of Canada and UK. 

The vast majority of the billionaires are married. Looking only to top 10, the pattern is the same.

In Brazil, the University of São Paulo and Federal University of Rio de Janeiro are the two universities which the billionaires studied the most. Both are public universities.

Between the brazillians, is possible to observe that the banking sector is the most frequent, followed by retail, hospital and beer.
