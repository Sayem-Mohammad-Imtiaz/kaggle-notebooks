### Questão 1 - Item C - Representação Gráfica
Para cada uma das variáveis, produza um ou mais gráficos, usando matplotlib, que descreva seu comportamento / caracteristica. Lembre-se que estes gráficos precisam ser compatíveis com a classificação da variável. 

```python
x = df[['uf','f_sup_79']].groupby('uf',as_index = False).sum().sort_values(by = 'f_sup_79', ascending = False)
x
#Eleitores por UF
fig = go.Figure(
    data=[go.Bar(x=x['uf'], y=x['f_sup_79'])],
    layout=go.Layout(
        title=go.layout.Title(text="Eleitores Sup 79 por estado")
    )
)

py.iplot(fig)  
```

```python
x = df[['uf','total_eleitores']].groupby('uf',as_index = False).sum().sort_values(by = 'total_eleitores', ascending = False)
x
#Eleitores por UF
fig = go.Figure(
    data=[go.Bar(x=x['uf'], y=x['total_eleitores'])],
    layout=go.Layout(
        title=go.layout.Title(text="Eleitores por estado")
    )
)

py.iplot(fig)  
```

```python
#Eleitores por gênero
masculino = df['gen_masculino'].sum()
feminino = df['gen_feminino'].sum()
n_if = df['gen_nao_informado'].sum()
langs = ['Masculino', 'Feminino', 'Não Informado']
Valores = [masculino, feminino, n_if]
trace = go.Pie(labels = langs, values = Valores)
data = [trace]
fig = go.Figure(data = data, layout=go.Layout(title=go.layout.Title(text="% de Eleitores Totais por Gênero")))
py.iplot(fig)
```

```python
#Municípios por estado
z = df[['uf','nome_municipio']].groupby('uf', as_index = False).count().sort_values(by = 'nome_municipio', ascending = False)
z

fig = go.Figure(
    data=[go.Bar(x=z['uf'], y=z['nome_municipio'])],
    layout=go.Layout(
        title=go.layout.Title(text="Quantidade de município por estado")
    )
)

py.iplot(fig)
```
