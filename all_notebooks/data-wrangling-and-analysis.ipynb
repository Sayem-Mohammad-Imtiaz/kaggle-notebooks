**By grouping we observed that the 2015 and 2016 year data of county allegany is distributed in two columns Allegany and Alleghany
so we must make them one.**

**Also in year 2015 and 2016 the 'Statewide Total' is named as 'New York State' we will make them same for future use**

```python
df[df['County']=='Alleghany']
```

```python
df[df['County']=='New York State']
```

```python
df.replace(to_replace=['Allegany','New York State'],value=['Alleghany','Statewide Total'],inplace=True)
```

```python
groupdf=df.groupby(['County'])
groupdf.describe()
```

**so now we are ready to visualize the county vise sales of Gallons**

```python
def plot_this(county):
    to_be_ploted=groupdf.get_group(county)
    plt.figure(figsize=(20,8))
    plt.title(to_be_ploted.iloc[0,1]+' County')
    plt.xlabel('Year')
    plt.ylabel('Estimated Sales, Thousands of Gallons')
    plt.plot(to_be_ploted['Year'],to_be_ploted['Estimated Sales, Thousands of Gallons'],linewidth=3,)
    plt.grid()
```

**Now to show the plot for any county we just need to call this function with the name of county**


```python
plot_this('Franklin')
plot_this('Yates')
```

```python
plot_this('Statewide Total')
```
