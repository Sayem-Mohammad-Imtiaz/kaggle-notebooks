```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../input/dados-de-suicdio-no-brasil-2014-a-2018/datasus_suicidio_2014_2018.csv', encoding='ISO-8859-1')
```

```python
df_suicidio = df.loc[df.CIRCOBITO == 'Suic√≠dio']
```

```python
df_suicidio = df_suicidio[['estado','ano','idade','SEXO', 'RACACOR', 'ESTCIV', 'LOCOCOR']]
```

# Suicides in Brazilian states

```python
df_suicidio.estado.value_counts().plot(kind='bar', figsize=(14,7), colormap='Set2', title='Suicides in Brazilian states')
```

# Distribution of suicides by age
The average age is 42 years

```python
plt.figure(figsize=(14,7))
plt.title('Distribution of suicides by age')
sns.histplot(data=df_suicidio.idade, kde=True, bins=50)
df_suicidio.idade.mean()
```

# Distribution of suicides by age and sex

```python
plt.figure(figsize=(14,7))
plt.title('Distribution of suicides by age and sex')
df_suicidio_m = df_suicidio.loc[df_suicidio.SEXO == 'Masculino']
df_suicidio_f = df_suicidio.loc[df_suicidio.SEXO == 'Feminino']
sns.kdeplot(data=df_suicidio_m.idade, shade=True)
sns.kdeplot(data=df_suicidio_f.idade, shade=True)
```

# Number of suicides by sex

```python
df_suicidio.SEXO.value_counts().plot(kind='barh', figsize=(10,5),colormap='Set2', title='Number of suicides by sex' )
```

# Number of suicides by marital status

```python
df_suicidio.ESTCIV.value_counts().plot(kind='barh', title='Number of suicides by marital status', figsize=(10,5), colormap='Set2')
```

```python
figure, axes = plt.subplots(1, 2, figsize=(14,7), gridspec_kw={'wspace': 0.4})
df_suicidio_m.ESTCIV.value_counts().plot(kind='barh', ax=axes[0])
axes[0].set_title('Masculino')
df_suicidio_f.ESTCIV.value_counts().plot(kind='barh', ax=axes[1],  colormap='Set2')
axes[1].set_title('Feminino')
figure.suptitle('Number of suicides by marital status and sex')
```

# Number of suicides per year

```python
plt.figure(figsize=(10,5))
plt.title('Number of suicides per year')
sns.lineplot(data=df_suicidio.ano)
```

# Number of suicides per establishment

```python
df_suicidio.LOCOCOR.value_counts().plot(kind='barh', colormap='Set2', figsize=(8,5),
                                       title='Number of suicides per establishment')
```

# Number of suicides by race / skin color

```python
df_suicidio.RACACOR.value_counts().plot(kind='barh', colormap='Set2', figsize=(10,5), title='Number of suicides by race / skin color')
```
