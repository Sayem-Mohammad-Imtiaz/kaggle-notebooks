Next we try to extract information from time stamp column

```python
#timeStamp column contains a lot of information which can be extracted.
type(df['timeStamp'][0])
```

```python
#timeStamp entries are a string. We have to convert it to datetime format

df['timeStamp']= pd.to_datetime(df['timeStamp'])
type(df['timeStamp'][0])
```

```python
#Create three new columns which contain information of day, date, hour
df['Date']=df['timeStamp'].apply(lambda x: x.date())
df['Day'] =df['timeStamp'].apply(lambda x:x.dayofweek)
df['Hour'] =df['timeStamp'].apply(lambda x:x.hour)
```

```python
df.head()
```

```python
# Day is in number, we want to map it to Monday to Sunaday so we use .map() function
getday={0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day']=df['Day'].map(getday)
```

```python
df.head()
```

Now with the extra information in these 3 columns we can extract lot of extra infomation, like which part of the day frequency of calls is more, etc

```python
plt.figure(figsize=(12,6))
df.groupby('Date')
df2=df.groupby('Date').count()
df2['twp'].plot.line()
```

```python
plt.figure(figsize=(12,6))
df3 = df[df['Reason'] == 'Traffic'].groupby('Date').count()
df3['twp'].plot.line()
```

On 2nd, 3rd and 7th of March 2018, number of 911 calls for traffic are very high

```python
plt.figure(figsize=(12,6))
df3 = df[df['Reason'] == 'Fire'].groupby('Date').count()
df3['twp'].plot.line()
```

Number of fire calls were very high on 2nd March 2018

```python
plt.figure(figsize=(12,6))
df3 = df[df['Reason'] == 'EMS'].groupby('Date').count()
df3['twp'].plot.line()
```

```python
dayHour = df.groupby(by=['Day','Hour']).count()['Reason'].unstack()
dayHour.head()
plt.figure(figsize=(12,6))
sns.heatmap(dayHour)
```

Majority of the 911 calls are between 7am and 7pm. But there are a substantial number of calls at night and early morning hours as well
