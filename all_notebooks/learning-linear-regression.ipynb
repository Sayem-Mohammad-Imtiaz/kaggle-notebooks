```python
df = pd.read_csv('/kaggle/input/salary-data-simple-linear-regression/Salary_Data.csv')
```

```python
df.info()
```

```python
df.rename(columns={'YearsExperience':'YOE', 'Salary':'Package'}, inplace=True)
```

There is no null value or missing value in dataset. so no need to do any data preprocessing here. We will check if any duplicate exists.

```python
print(df.shape)
df.drop_duplicates(inplace=True) # By marking inplace = true , it will modify the df dataset itslef
print(df.shape)
```

Since the shape of dataset is not changed so it means there were no duplicates.
Now Its time to Divide the dataset into training dataset and test dataset.

```python
from sklearn.model_selection import train_test_split
x = df['YOE'].values
y = df['Package'].values
#splitting the data into Training set and Test set 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train.reshape(-1,1), y_train)
```

```python
y_pred=model.predict(x_test.reshape(-1,1))
print(y_pred)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_train,y_train,c="red",lw=0.5)
plt.plot(x_train,model.predict(x_train.reshape(-1,1)),c="blue",lw=0.5)
plt.title("Salary Vs Experience (Training set)")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.show()
```
