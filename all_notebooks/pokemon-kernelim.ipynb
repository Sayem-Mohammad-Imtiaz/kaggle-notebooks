### C. NESTED FUNCTION
* Function inside function

```python
def square():
    """return square of value"""
    def add():
        """add two local variables"""
        x = 2
        y = 3
        z = x + y
        return z
    return add()**2
print(square())
```

### D. DEFAULT AND FLEXIBLE VARIABLES
* Default argument example:
<br> def (a, b=1):
        b=1 is default argument if it's not given

* Flexible argument example:
<br> def(*args):
        *args can be one or more
<br>def f(**kwargs):
        **kwargs is a dictionary

```python
#default arguments
def f(a, b=1, c=2):
    y = a + b + c
    return y
print(f(5))
# or
print(f(5, 4, 3))
```

```python
# flexible arguments
def f(*args):
    for i in args:
        print(i)
f(1)
print('')
f(1,5,3)
```

```python
# flexible arguments **kwargs that is dictionary
def f(**kwargs):
    """print key and value of dict"""
    for key,value in kwargs.items():
        print(key, ":", value)
f(country = 'spain', capital = 'madrid', population = 123456)
```

### E. LAMBDA FUNCTION
Faster way of writing function

```python
#lambda function
square = lambda x: x**2  # x is the argument
print(square(4))

tot = lambda x,y,z: x+y+z # x,y,z are the arguments
print(tot(1,2,3))
```

### F. ANONYMOUS FUNCTION
It is like the Lambda function but it can take more than one argument.

* map(func, list): applies that specific function to all items in the list
                    and returns a list back.

```python
num_list = [1,2,3]
y = map(lambda x:x**2, num_list)
print(list(y))
```

### G. ITERATORS
* Iterable is an object that can return an iterator back.
* list, strings and dictionaries are iterable.
* Iterables are generally objects that we can call a loop on them.

```python
# iteration example

name = 'ronaldo' # string can be made an iterable object.
it = iter(name) # make string an iterable object.
print(next(it)) # print next iteration
print(*it)      # print remaining iteration
```

```python
# zip example zip(list)

list1 = [1,2,3,4]
list2 = [5,6,7,8]
num_label = ["num", "numin"]
listcol = [list1, list2]
z = zip(num_label, listcol)
print(z) # z is an obejct.

z_list = list(z) # z_list is a list.
print(z_list)

z_dict = dict(z_list)
print(z_dict)

dfz = pd.DataFrame(z_dict)
dfz
```

```python
# unzip example zip(*list)

un_zip = zip(*z_list) # un_zip is an object.
un_list1, un_list2 = list(un_zip) # un_listx is a tuple.
print(un_list1)
print(un_list2)
print(type(un_list2))
print(type(list(un_list2)))
```

### B. BUILDING DATAFRAMES FROM SCRATCH
* We can build dataframes ourselves instead of CSV files.
* 1. We make two 1 column arrays which is called a series.
* 2. We label the columns of those. So we make another series on which we are going to put our labels.
* 3. * We put our columns (which are series) together as list_col = [series1, series2] 
* 3. * We zip the label series and the new list together to make another list. (zip is an object.)
* 3. * list(zip(list_label, list_col))
* 4. We convert that list into a dictionary.
* 5. Finally we convert dict to a dataframe.


```python
country = ["Spain", "France"]
population = ["11", "12"]
list_label = ["country", "population"]
list_col = [country, population]
zipped = list(zip(list_label, list_col))
data_dict = dict(zipped)
dfz = pd.DataFrame(data_dict)
dfz
```

```python
# We can add new columns to that dataframe
dfz["capital"] = ["madrid", "paris"]
dfz
```

```python
# We can broadcast a value to all rows in a column
dfz["income"] = 0
dfz
```

### F. RESAMPLING PANDAS TIME SERIES
* Resampling is a method performed over different time intervals. "M" = month, "A" = year
* Downsampling : decrease the datetime freq. from daily to weekly or monthly
* Upsampling   : increase the datetime freq. from monthly to weekly or daily
* Interpolate  : put values in a time interval with interpolation.

```python
data2.resample("A").mean()
# We take the mean(average) of years.
```

```python
data2.resample("M").mean()
# We take the mean of every month in every year.
# There are NaN values since there are no values for those months.
```

```python
data2.resample("M").max().interpolate("linear")
# used .max() instead of .first()
# .first() causes an error "Invalid fill method" for interpolate().
```

```python
data2.resample("M").mean().interpolate("linear")
```
