```python
df=pd.read_csv('/kaggle/input/premier-league-players-defense-stats/PL_Players.csv')
df.info()
```

# We will extract those player who have played on Mid Field Position 

```python
mfPlayers_df=df[df['Pos'].str.contains(pat='MF')]
mfPlayers_df.set_index('Player',inplace=True)
mfPlayers_df
```

# Now we will sort the Players according to their:
# 1. Tacles won in Mid field 
# 2. Press in midfield
# 3. Blocks
# 4. Passes Blocked

```python
mfPlayers_df.sort_values(by=['Mid 3rd','Mid 3rd.1','Blocks','Pass'],ascending=False,inplace=True)
mfPlayers_df
```

# Prayer Personal Stats For Mid field

```python
def player_mid_field_stats(player):
    plt.figure(figsize=(15,10))
    plt.title(label=player+" Mid Field Stats",fontsize=20)
    plt.pie(mfPlayers_df.loc[player,['Mid 3rd','Mid 3rd.1','Blocks','Pass']],labels=['Tackles won in midfield','Press in midfield','Blocks','Passes Blocked'])
    plt.show()
```

```python
player_mid_field_stats('John McGinn')
player_mid_field_stats('Oriol Romeu')
```

# So The Manager's Top 10 options for a Mid field player are:

```python
top10_df=mfPlayers_df.head(10)
top10_df
```

**Selecting only useful columns**

```python
top10_df.drop(df.columns.difference(['Mid 3rd','Mid 3rd.1','Blocks','Pass','Player']),axis=1,inplace=True)
print(top10_df)
```

**Now we add these value to get our player total stats**

```python
top10_df=top10_df.sum(axis=1)
```

```python
color=['#d10d2f','#d52342','#d93954','#dd4e67','#e2647a','#e67a8d','#ea909f','#eea5b2','#f2bbc5','#f6d1d7']
color.reverse()
top10_df.sort_values(inplace=True)
plt.figure(figsize=(15,10))
plt.title('# Top 10 Mid Field Players',fontsize=30)
plt.xlabel("player's total points")
plt.barh(top10_df.index,top10_df,color=color)
plt.yticks(fontsize=20)
plt.show()
```
