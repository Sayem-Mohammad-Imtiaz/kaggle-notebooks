```python
#Codes by https://www.kaggle.com/muhammadismail99/data-analysis-of-guf-pub-dataset/notebook

sns.countplot(data = df, x = 'HS')
plt.title('Home Score')
```

```python
df["AS"].value_counts()
```

```python
sns.countplot(data = df, x = 'AS', color='b')
plt.title('Away Score')
```

```python
home=df['HS'].value_counts().sort_index()
data = [go.Pie(
        labels = home.index,
        values = home.values,
        title = 'Home Scores',    
        hoverinfo = 'label+value',
)]
plotly.offline.iplot(data, filename='HS')
```

```python
away=df['AS'].value_counts().sort_index()
data = [go.Pie(
        labels = away.index,
        values = away.values,
        title = 'Away Scores',    
        hoverinfo = 'label+value',
)]
plotly.offline.iplot(data, filename='AS')
```
