```python
data = pd.read_csv('../input/income-adult/adult_data.csv',header=0)
data.head()
```

```python
test_data=pd.read_csv('../input/income-adult/adult_test.csv')
test_data.info()
```

```python
lis=[' education', ' salary']
data[lis].value_counts()
```

```python
data["Salary"] = np.where(data[" salary"].str.contains(">"), 1, 0)
test_data["Salary"] = np.where(test_data[" salary"].str.contains(">"), 1, 0)
```

```python
data.info()
```

```python
from sklearn import preprocessing
  
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
data[' workclass']= label_encoder.fit_transform(data[' workclass'])
data[' education']= label_encoder.fit_transform(data[' education'])
data[' occupation']= label_encoder.fit_transform(data[' occupation'])
data[' native-country']= label_encoder.fit_transform(data[' native-country'])

#test data

test_data[' workclass']= label_encoder.fit_transform(test_data[' workclass'])
test_data[' education']= label_encoder.fit_transform(test_data[' education'])
test_data[' occupation']= label_encoder.fit_transform(test_data[' occupation'])
test_data[' native-country']= label_encoder.fit_transform(test_data[' native-country'])

imp_cols=[' workclass',' education', ' occupation', ' native-country', 'age', ' hours-per-week']



```

```python
X_train=data[imp_cols]
y_train=data['Salary']
X_test=test_data[imp_cols]
y_test=test_data['Salary']
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(max_depth=7,random_state=42)
rfc.fit(X_train,y_train)
```

```python
rfc.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,rfc.predict(X_test))
```
