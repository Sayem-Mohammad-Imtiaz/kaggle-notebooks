```python
df_html = pd.read_csv(files[0],parse_dates=["Closing Date"])
print(df_html)
```

```python
df_html.info()
```

```python
df_html.columns

df_html.drop([df_html.columns[0]],axis=1,inplace=True)

df_html.head(5)
```

```python
count_by_city = df_html.groupby(['City']).agg(np.size)

count_by_city.columns

count_by_city.drop(count_by_city.columns[1:],axis=1,inplace=True)
count_by_city.rename(columns = {count_by_city.columns[0]: 'count_by_cities'}, inplace = True)

count_by_city
```

```python
count_by_aqu = df_html.groupby(['Acquiring Institution']).agg(np.size)

count_by_aqu.columns

count_by_aqu.drop(count_by_aqu.columns[1:],axis=1,inplace=True)
count_by_aqu.rename(columns = {count_by_aqu.columns[0]: 'count_by_aq_bank'}, inplace = True)

count_by_aqu
```

```python
count_by_banks = df_html.groupby(['Bank Name']).agg(np.size)

count_by_banks
```

```python
count_by_closing_year = df_html.groupby(['Year']).agg(np.size)

count_by_closing_year.columns

count_by_closing_year.drop(count_by_closing_year.columns[1:],axis=1,inplace=True)
count_by_closing_year.rename(columns = {count_by_closing_year.columns[0]: 'count_by_closing_year'}, inplace = True)

count_by_closing_year
```

```python
count_by_state = df_html.groupby(['State']).agg(np.size)

count_by_state.columns

count_by_state.drop(count_by_state.columns[1:],axis=1,inplace=True)
count_by_state.rename(columns = {count_by_state.columns[0]: 'count_by_states'}, inplace = True)

#count_by_state
print("number of states:",len(count_by_state))
print(count_by_state)
#type(count_by_state)
```
