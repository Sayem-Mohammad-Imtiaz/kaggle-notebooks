## DATA PROCESS

#### TRANSFORMATION

```python
IMAGE_LIST = []
MASK_LIST = []

for turtle_i,mask_i in zip(JPG_Series,JPG_Series):
    
    Reading_Image = reading_image(turtle_i)
    Canny_Image = reading_canny(mask_i)
    Reading_Image_2D = Reading_Image[:,:,0]

    Blend_Image = cv2.addWeighted(Reading_Image_2D,0.8,Canny_Image,0.4,0.5)
    
    Resized_Main = cv2.resize(Reading_Image,(180,180))
    Resized_Mask = cv2.resize(Blend_Image,(180,180))
    
    IMAGE_LIST.append(Resized_Main)
    MASK_LIST.append(Resized_Mask)
```

```python
Early_Stopper = tf.keras.callbacks.EarlyStopping(monitor="loss",patience=3,mode="min")
Checkpoint_Model = tf.keras.callbacks.ModelCheckpoint(monitor="val_accuracy",
                                                      save_best_only=True,
                                                      save_weights_only=True,
                                                      filepath="./modelcheck")
```

#### PARAMETERS

```python
compile_loss = "binary_crossentropy"
compile_optimizer = Adam(lr=0.00001)
output_class = 1
```

#### STRUCTURE

```python
Encoder_AE = Sequential()
Encoder_AE.add(Conv2D(32,(5,5),kernel_initializer = 'he_normal'))
Encoder_AE.add(BatchNormalization())
Encoder_AE.add(ReLU())
#
Encoder_AE.add(Conv2D(64,(5,5),kernel_initializer = 'he_normal'))
Encoder_AE.add(BatchNormalization())
Encoder_AE.add(ReLU())
#
Encoder_AE.add(Conv2D(128,(5,5),kernel_initializer = 'he_normal'))
Encoder_AE.add(BatchNormalization())
Encoder_AE.add(ReLU())


Decoder_AE = Sequential()
Decoder_AE.add(Conv2DTranspose(64,(5,5)))
Decoder_AE.add(ReLU())
#
Decoder_AE.add(Conv2DTranspose(32,(5,5)))
Decoder_AE.add(ReLU())
#
Decoder_AE.add(Conv2DTranspose(output_class,(5,5)))
Decoder_AE.add(ReLU())
```

```python
Auto_Encoder = Sequential([Encoder_AE,Decoder_AE])
Auto_Encoder.compile(loss=compile_loss,optimizer=compile_optimizer)
```

```python
Model_AutoEncoder = Auto_Encoder.fit(X_Train,X_Test,epochs=10,callbacks=[Checkpoint_Model])
```
