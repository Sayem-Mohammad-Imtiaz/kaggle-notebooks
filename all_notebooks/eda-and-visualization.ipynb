```python
#Importing data
df=pd.read_csv("../input/911.csv")
df.head()
```

```python
#datatypes of the various columns in the  dataframe
df.dtypes
```

   **Handling Missing values**

```python
#checking for the null values in the column
print(df.isnull().sum())
#checking for the percentage of null values in the column
print(100*(df.isnull().sum()/326425))
#we observe that zip has 12.12% missing values and twp has 0.035% missing values
```

```python
#finding duplicates
df = df[~df.index.duplicated(keep='first')]
#we see that no duplicate values are present in the rows
```

**Total number of calls based on their types**

```python
#Creating 'type' column based on 'title' to separate the 'type' part from 'title'
df['type']=df['title'].apply(lambda x: x.split(':')[0])
df.head()
#counting the values of 'type' column  
df1=df['type'].value_counts()
df1
#plot to show the number of types
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
df1.plot(kind='bar')
plt.show() 
#EMS is the highest type(category) for Emergency 911 calls
```

**Top 5 townships based on total calls **

```python
#creating crosstabs based on 'type' column
tab=pd.crosstab(index=df['twp'], columns=df['type'])
#creating a column named 'totalcalls' 
tab['totalcalls']=tab['EMS']+tab['Fire']+tab['Traffic']
print(tab.head())
#computing top 5 townships
top5=tab.nlargest(5, 'totalcalls', keep='first')
top5
# graphically
sns.set()
top5['totalcalls'].plot(kind='bar')
plt.show()
#Lower Merion is the highest based on total calls
```

**Top 5 townships for 'EMS' Emergency calls**

```python
#calculating top5 EMS calls 
top5_EMS=tab.nlargest(5,'EMS').reset_index()
top5_EMS
#graphically
sns.set()
sns.barplot(x='EMS',y='twp',data=top5_EMS)
plt.show()
#NORRISTOWN has the highest EMS emergency calls
```

**Top 5 townships for 'Fire' Emergency calls**

```python
#calculating top5 fire calls 
top5_Fire=tab.nlargest(5,'Fire').reset_index()
top5_Fire
#graphically
sns.set()
sns.barplot(x='Fire',y='twp',data=top5_Fire)
plt.show()
#LOWER MERION has the highest Fire emergency calls
```

**Top 5 townships for 'Traffic' Emergency calls**

```python
#calculating top5 Traffic calls 
top5_Traffic=tab.nlargest(5,'Traffic').reset_index()
top5_Traffic
#graphically
sns.set()
sns.barplot(x='Traffic',y='twp',data=top5_Traffic)
plt.show()
#LOWER MERION has the highest Traffic emergency calls
```

```python
#locating the type of timestamp
type(df['timeStamp'].iloc[0])
#Converting the timestamp column into datetime format from str format 
df['timeStamp'] = pd.to_datetime(df['timeStamp'])
#creating new columns-hour,minutes,year,month,day from 'timestamp' column
df['Hour'] = df['timeStamp'].apply(lambda x: x.hour)
df['Minutes'] = df['timeStamp'].apply(lambda x: x.minute)
df['Year'] = df['timeStamp'].apply(lambda x: x.year)
df['Month'] = df['timeStamp'].apply(lambda x: x.month)
df['Dayofweek'] = df['timeStamp'].apply(lambda x: x.dayofweek)
```

```python
#using dmap to map string values to integer values since day appears as integer values from 0-6
dayofweek= {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Dayofweek']= df['Dayofweek'].map(dayofweek)
#countplot to see number of emergency calls on each day of the week
sns.countplot(x='Dayofweek',data=df,hue='type')
#title
plt.title('Day wise countplot for different types')
#relocate the legends
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
#countplot to see number of emergency calls on each year 
sns.countplot(x='Year',data=df,hue='type')
#title
plt.title('Year wise countplot for different types')
#relocate the legends
plt.legend(bbox_to_anchor=(1.05, 0), loc=2, borderaxespad=0.)
plt.show()
```

```python
#using dmap to map string values to integer values since month appears as integer values frpom 1-12
month= {1:'Jan',2:'Feb',3:'Mar',4:'April',5:'May',6:'June',7:'July',8:'Aug',9:'Sept',10:'Oct',11:'Nov',12:'Dec'}
df['Month']= df['Month'].map(month)
#countplot to see number of emergency calls on each month of the year
sns.countplot(x='Month',data=df,hue='type')
#title
plt.title('Monthwise countplot for different types')
#relocate the legends
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

**Heatmaps for all 911 calls**

```python
#pivot table for month and hour from df and calculating the sum
calls=df.pivot_table(values='e',index='Month',columns='Hour',aggfunc=np.sum)
calls
#heatmaps
plt.figure(figsize=(14,7))
sns.heatmap(calls)
plt.show()
#Heat map shows the intensity of calls during the month and time period of the day
#It can be inferred that there are more calls in the month of January,March and December especially during the working hours.
```
