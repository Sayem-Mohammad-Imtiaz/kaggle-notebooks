```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import plotly.express as px
```

## **Data Wrangling**

```python
data_set = pd.read_csv("/kaggle/input/marketing-data/marketing_data.csv")
data_set.head(5)
```

'Income' columnn has 24 Null values

### Removing the Null values and Filling them with the Mean values instead

```python
data_set[' Income '] = data_set[' Income '].str.replace('$', '')
data_set[' Income '] = data_set[' Income '].str.replace(',', '')
data_set.head(5)
```

```python
data_set[' Income '] = data_set[' Income '].astype(float)
data_set[' Income '].fillna((data_set[' Income '].mean()), inplace = True)
```

```python
data_set.head(5)
```

```python
data_set.isnull().sum()
```

```python
fig, (axes_1,axes_2) = plt.subplots(1,2,figsize=(20,10))

labels = data_set['Kidhome'].unique()
values = data_set['Kidhome'].value_counts()
axes_1.pie(values, labels = labels, autopct = '%1.1f%%') #plot first pie


labels = data_set['Teenhome'].unique()
values = data_set['Teenhome'].value_counts()
axes_2.pie(values, labels = labels, autopct = '%1.1f%%') #plot second pie

axes_1.set(aspect="equal", title='With kids')
axes_2.set(aspect="equal", title='With teens')
plt.show()
```
