```python
clasi=[RandomForestClassifier(), DecisionTreeClassifier(),SVC(), GradientBoostingClassifier(), AdaBoostClassifier(), KNeighborsClassifier(3), GaussianNB() ]
```

```python
L={}
for x in clasi:
    model=x.__class__.__name__
    result=cross_val_score(x, df.iloc[:,:-1], df.iloc[:,-1], cv=4,scoring='accuracy')
    L[model]=result.mean()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.xticks(rotation=90)
sns.barplot(list(L.keys()), list(L.values()))
```
