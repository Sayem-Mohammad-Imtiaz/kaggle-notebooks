```python
aug_train = pd.read_csv('../input/hr-analytics-job-change-of-data-scientists/aug_train.csv')
aug_test = pd.read_csv('../input/hr-analytics-job-change-of-data-scientists/aug_train.csv')
aug_train.head()
```

```python
aug_train = aug_train.dropna(axis = 0)
```

```python
aug_train.describe()
```

```python
target = aug_train['target'].value_counts()
print(target)

fig, ax = plt.subplots(figsize=(15,5))
colors=['darkgrey','lightgray']
ax.pie(target , colors=colors, autopct='%.2f%%')
ax.set_title('Target')
ax.legend(['No','Yes'], title = 'Target', bbox_to_anchor=(1,1))
```

```python
fig, ax = plt.subplots(figsize=(15,5))
sns.countplot(ax=ax, data = aug_train, x='city', order = aug_train['city'].value_counts().head(10).index)
ax.set_title('10 Cities Origin of Most Candidates')
ax.set_ylabel('Candidates')
ax.set_xlabel('City')
```

```python
cdi = aug_train.drop_duplicates(subset=['city']).sort_values(by='city_development_index', ascending=False).head(20)
print("Lowest CDI: ", aug_train.city_development_index.min())
print("Highest CDI: ", aug_train.city_development_index.max())
fig, ax = plt.subplots(figsize=(15,5))
sns.barplot(data = cdi, x='city_development_index', y='city')
ax.set_title('Top 20 Highest City Development Index')
ax.set_ylabel('City')
ax.set_xlabel('CDI')
```

```python
#aug_train_1 = aug_train.loc[aug_train['target'] == 1]
gender = aug_train['gender'].value_counts()
print(gender)

fig, ax = plt.subplots(figsize=(15,5))
colors=['pink','hotpink','deeppink']
ax.pie(gender, colors=colors, autopct='%.2f%%')
ax.set_title('Candidate Gender')
ax.legend(['Male','Female','Other'], title = 'Gender', bbox_to_anchor=(1,1))
```

```python
re = aug_train['relevent_experience'].value_counts()
print(re)

fig, ax = plt.subplots(figsize=(15,5))
colors=['cyan','darkturquoise']
ax.pie(re , colors=colors, autopct='%.2f%%')
ax.set_title('Relevent Experience')
ax.legend(['Has','No'], title = 'Relevent Experience', bbox_to_anchor=(1,1))
```

```python
enroll_univ = aug_train['enrolled_university'].value_counts()
print(enroll_univ)

fig, ax = plt.subplots(figsize=(15,5))
colors=['greenyellow','chartreuse','palegreen']
ax.pie(enroll_univ, colors=colors, autopct='%.2f%%')
ax.set_title('Enrolled University')
ax.legend(['No Enrollment', 'Full time course', 'Part time course'], 
          title = 'Enrolled', 
          bbox_to_anchor=(1,1))
```

```python
edu_lv = aug_train['education_level'].value_counts()
print(edu_lv)

fig, ax = plt.subplots(figsize=(15,5))
colors=['darksalmon','coral','orangered']
ax.pie(edu_lv, colors=colors, autopct='%.2f%%')
ax.set_title('Education Level')
ax.legend(['Graduate', 'Master', 'Phd'], 
          title = 'Edu Lv', 
          bbox_to_anchor=(1,1))
```
