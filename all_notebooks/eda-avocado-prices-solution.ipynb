**Compare the features 'PLU4046','PLU4225','PLU4770','AveragePrice' using pairplot**

```python
sns.pairplot(data=data[['PLU4046','PLU4225','PLU4770','AveragePrice']],height=3, kind='scatter')
```

<font color='blue'>**OBSERVATION :** PLU(Price LookUP codes) are inversely proportional to Average price</font>

**Compare the features 'PLU4046','PLU4225','PLU4770','AveragePrice' using pairplot**

```python
sns.pairplot(data=data[['PLU4046','AveragePrice']],height=4, kind='scatter')
```

<font color=blue>**OBSERVATION :**The price of Avocados range from 0.5 to 2.3 with peak between 1 to 1.5 </font> 

**Compare the features 'Total Volume','Total Bags','AveragePrice' using pairplot**

```python
sns.pairplot(data=data[['Total Volume','Total Bags','AveragePrice']],height=4, kind='scatter')
```

<font color=blue>**OBSERVATION :**Total volumes is directly proportional to Total bags </font> 

**Compare the features 'Small Bags','Large Bags','AveragePrice' using pairplot**

```python
sns.pairplot(data=data[['Small Bags','Large Bags','AveragePrice']],height=4, kind='scatter')
```

<font color=blue>**OBSERVATION :** No. of large and small bags is inversely proportional to the Average price</font> 

**Compare the features 'XLarge Bags','year','AveragePrice' using pairplot**

```python
sns.pairplot(data=data[['XLarge Bags','year','AveragePrice']],height=4, kind='scatter')
```

<font color=blue>**OBSERVATION :** No. of extra large bages is inversely proportional to Average price</font>
