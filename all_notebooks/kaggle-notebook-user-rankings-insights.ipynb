```python
def plot_distribution(lists,df,field_name):
    for l in lists:
        print("List =>",l)
        klib.dist_plot(df[df[field_name]==l])
    return None
```

```python
df['Tier'].unique().tolist()
```

```python
plot_distribution(df['Tier'].unique().tolist(),df,'Tier')
```

```python
rnkg = (df.groupby(['Tier']).agg({'UserId':'count'}).rename(columns={'UserId':'Ranking-Percentage'})/df.shape[0]*100).reset_index()
```

```python
rnkg['Tier_desc']=rnkg['Tier'].map({2:'Expert',3:'Master',4:'GrandMaster'})
```

```python
source = df['Tier_desc'].value_counts()
```

```python
fig = go.Figure(data=[go.Pie(labels=source.index,values=source.values)])
fig.update_layout(title='Kaggle Title distribution')
fig.show()
```

```python
plt.figure(figsize=(35,35))
fig = go.Figure()
for idx, values in enumerate([df['TotalGold'].values.tolist(), df['TotalSilver'].values.tolist(), df['TotalBronze'].values.tolist()]):
    if idx == 0:
        color = "Red"
    if idx == 1:
        color = "Green"
    if idx == 2:
        color = "Blue"
    fig.add_trace(go.Box(x=[color]*len(values), y=values, name=color, marker=dict(color=color.lower())))
    
fig.update_layout(yaxis_title="Medals Count", xaxis_title="Kaggle Medals",
                  title="Medals Box plot", template="plotly_white")
```

```python
fig = ff.create_distplot([df['TotalGold'].values.tolist()], group_labels=["R"], colors=["red"])
fig.update_layout(showlegend=False, template="simple_white")
fig.update_layout(title_text="Distribution of Gold Medels")
fig.data[0].marker.line.color = 'rgb(0, 0, 0)'
fig.data[0].marker.line.width = 0.5
fig
```

```python
fig = ff.create_distplot([df['TotalSilver'].values.tolist()], group_labels=["G"], colors=["green"])
fig.update_layout(showlegend=False, template="simple_white")
fig.update_layout(title_text="Distribution of Kaggle Silver Medels")
fig.data[0].marker.line.color = 'rgb(0, 0, 0)'
fig.data[0].marker.line.width = 0.5
fig
```

```python
fig = ff.create_distplot([df['TotalBronze'].values.tolist()], group_labels=["B"], colors=["blue"])
fig.update_layout(showlegend=False, template="simple_white")
fig.update_layout(title_text="Distribution of Kaggle Bronze medals")
fig.data[0].marker.line.color = 'rgb(0, 0, 0)'
fig.data[0].marker.line.width = 0.5
fig
```

## 3D Scatter plot of Kaggle Gold , Silver & Bronze medals
