```python
X=df.iloc[:,1:18]
Y=df.iloc[:,19]
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
def model_fun (X,Y):
    model=LogisticRegression(solver='liblinear')
    model.fit(X_train,y_train)
    prediction=model.predict(X_test)
    print("confusion_matrix:",confusion_matrix(y_test, prediction)),
    print("accuracy_score :",accuracy_score(y_test, prediction))
    print("log_loss :",log_loss(y_test, prediction))
    print("mean_squared_error :",mean_squared_error(y_test, prediction))
    print("f1_score :",f1_score(y_test, prediction))
    return accuracy_score(y_test, prediction) 
```
