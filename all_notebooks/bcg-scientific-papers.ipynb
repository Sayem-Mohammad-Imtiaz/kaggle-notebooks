```python
import time
import warnings
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

warnings.filterwarnings('ignore')
```

```python
# load metadata
t1 = time.time()
df = pd.read_csv('../input/hackathon/BCG_COVID-19_scientific_papers.csv')
t2 = time.time()
print('Elapsed time:', t2-t1)
```

```python
# define keyword
my_keyword = 'BCG'
```

```python
def word_finder(i_word, i_text):
    found = (str(i_text).lower()).find(str(i_word).lower()) # avoid case sensitivity
    if found == -1:
        result = 0
    else:
        result = 1
    return result

# partial function for mapping
word_indicator_partial = lambda text: word_finder(my_keyword, text)
# build indicator vector (0/1) of hits
keyword_indicator = np.asarray(list(map(word_indicator_partial, df.Summary)))
```

```python
# number of hits
print('Number of hits for keyword <', my_keyword, '> : ', keyword_indicator.sum())
```

```python
# add index vector as additional column
df['selection'] = keyword_indicator

# select only hits from data frame
df_hits = df[df['selection']==1]
```

```python
# show results
df_hits
```

```python
# show all abstracts
n = df_hits.shape[0]
for i in range(0,n):
    print(df_hits.Title.iloc[i],":\n")
    print(df_hits.Summary.iloc[i])
    print('\n')
```

```python
# make available for download
df_hits.to_csv('hits.csv')
```

#Test the BCG-Covid hypothesis, By Sruthijith K K & Ashish Kamat

Among the variables so far been proposed as correlating with the varying incidence and mortality of Covid-19, such as temperature, GDP and testing, the most promising has been the correlation with the BCGNSE -3.77 % (Bacillus Calmette-Guérin) vaccine, a century-old protector against tuberculosis (TB).

In a March 2020 paper, ‘BCG vaccination may be protective against Covid-19’,(bit.ly/3aXYBTM) researchers found that countries with aBCG vaccination policy saw 10 times fewer incidence and mortality of Covid-19, compared with those that didn’t have a BCG vaccination policy. The heatmap of Covid-19 incidence in Europe correlates neatly with the map of the countries there without a BCG vaccination policy.

Rich countries eradicated TB long before others, and haven’t had to inoculate their population with BCG. India, historically having a huge TB burden, has had a universal BCG vaccination policy for a long time.

There can be many confounding factors, as the paper acknowledges, and correlation doesn’t imply causation. But if results of definitive clinical trials confirm causation, BCG could become our armour against Covid, as it will likely become a seasonal affliction.

There’s an easy way of probing the correlation deeper. BCG, given at birth in India, doesn’t confer lifelong immunity. Whether the person still enjoys immunity from the vaccine or not can be determined using a skin test — PPD (purified protein derivative) test, or the Mantoux test, both reasonable options to the more expensive Igra (interferon gamma release assay) test. A positive test indicates a person still had BCGconferred immunity against TB.

If scientists can glean a valuable signal about the BCG-Covid-19 correlation using the PPD test, it could greatly aid policymakers faced with impossible trade-offs.
