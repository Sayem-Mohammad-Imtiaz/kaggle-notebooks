```python
df = pd.read_csv('../input/cusersmarildownloadsratioscsv/ratios.csv', delimiter=';', encoding = "utf8")
df.head()
```

```python
df.isnull().sum()
```

```python
#word cloud
from wordcloud import WordCloud, ImageColorGenerator
text = " ".join(str(each) for each in df.key_ratios)
# Create and generate a word cloud image:
wordcloud = WordCloud(max_words=200,colormap='Set3', background_color="black").generate(text)
plt.figure(figsize=(10,6))
plt.figure(figsize=(15,10))
# Display the generated image:
plt.imshow(wordcloud, interpolation='Bilinear')
plt.axis("off")
plt.figure(1,figsize=(12, 12))
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

#fill in mean for floats
for c in df.columns:
    if df[c].dtype=='float16' or  df[c].dtype=='float32' or  df[c].dtype=='float64':
        df[c].fillna(df[c].mean())

#fill in -999 for categoricals
df = df.fillna(-999)
# Label Encoding
for f in df.columns:
    if df[f].dtype=='object': 
        lbl = LabelEncoder()
        lbl.fit(list(df[f].values))
        df[f] = lbl.transform(list(df[f].values))
        
print('Labelling done.')
```

#Let's visualize Key ratios over years 2011-2020.

```python
#Codes by Tawej https://www.kaggle.com/tawejssh/covid-19

#Let's visualise Key ratios
ratio = df.groupby('key_ratios').sum()[['ratios_2011_09','ratios_2012_09','ratios_2013_09', 'ratios_2014_09', 'ratios_2015_09', 'ratios_2016_09', 'ratios_2017_09', 'ratios_2018_09', 'ratios_2019_09', 'ratios_2020_09', 'ttm' ]]
#evolution['Expiration Rate'] = (evolution['Expired'] / evolution['Cumulative']) * 100
#evolution['Discharging Rate'] = (evolution['Discharged'] / evolution['Cumulative']) * 100
ratio.head()
```

#To avoid a certain clumsy visualization here I separated 2011 till 2016

```python
#Codes by Tawej https://www.kaggle.com/tawejssh/covid-19

plt.figure(figsize=(20,7))
plt.plot(ratio['ratios_2011_09'], label='ratios_2011_09')
plt.plot(ratio['ratios_2012_09'], label='ratios_2012_09')
plt.plot(ratio['ratios_2013_09'], label='ratios_2013_09')
plt.plot(ratio['ratios_2014_09'], label='ratios_2014_09')
plt.plot(ratio['ratios_2015_09'], label='ratios_2015_09')
plt.plot(ratio['ratios_2016_09'], label='ratios_2016_09')
plt.legend()
#plt.grid()
plt.title('Key Ratios 2011 - 2016')
plt.xticks(ratio.index,rotation=45)
plt.xlabel('Key Ratios')
plt.ylabel('Count')
plt.show()
```

#Key ratios between 2017 - 2020

#Trailing 12 months (TTM):  

"It's the term for the data from the past 12 consecutive months used for reporting financial figures. A company's trailing 12 months represent its financial performance for a 12-month period; it does not typically represent a fiscal-year ending period."

https://www.investopedia.com/terms/t/ttm.asp#:~:text=Trailing%2012%20months%20(TTM)%20is%20the%20term%20for%20the%20data,a%20fiscal%2Dyear%20ending%20period.

```python
plt.style.use('ggplot')

#What about the ratio of Trailing 12 months (TTM)?
plt.figure(figsize=(20,7))
plt.plot(ratio['ttm'], label='Trailing 12 months (TTM)')

plt.legend()
plt.grid()
plt.title('TTM 2011-2020 ')
plt.xticks(ratio.index,rotation=45)
plt.ylabel('Rate %')
plt.show()
```
