```python
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
T = netflix_table[netflix_table.country == "Thailand"]
Thailand_chart_1 = sns.jointplot(data=T, x="release_year", y="rating")
```

```python
Thailand_chart_2 = sns.distplot(T.release_year, bins=15)
```

```python
Thailand_chart_3 = plt.hist(T.release_year, bins=15)
```

```python
sns.set_theme(style="ticks", palette="cool")
Thailand_chart_vio = sns.violinplot(data=T, x="release_year", y="rating", hue="type")
```

```python
Thailand_chart_box = sns.boxplot(data=T, x="release_year", y="rating", hue="type")
```

```python
C = netflix_table[netflix_table.country == "China"]
China_chart_boxbox = sns.boxplot(data=C, x="release_year", y="rating", hue="type")
```
