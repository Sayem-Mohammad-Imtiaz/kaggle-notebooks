## Reading Data

```python
df=pd.read_csv("/kaggle/input/coursera-ipo-tweets/CourseraIPOTweets.csv")
df.head()
```

### Rows and columns 
- The size of data is 5 rows and 36 Columns

```python
df.shape
```

### Checking for Null values in each columns

```python
df.isnull().sum()
```

### Droping Null values
- The Columns have been droped which where having more numbers of null values.

```python
df.drop(["place","quote_url","thumbnail","near","geo","source","user_rt_id","user_rt","retweet_id","retweet_date","translate","trans_src","trans_dest"],axis=1,inplace=True)
```

### Checking Numbers of zeros in each column.

```python
(df==0).sum()
```

### Droping Columns
- the columns have been droped which where having more number of zeros present in it.

```python
df.drop(["timezone","replies_count","retweets_count","retweet"],axis=1,inplace=True)
```

```python
df.head()
```

- After dropping columns which contains null values and zeros.
- The final size of data is 5 rows and 19 columns
