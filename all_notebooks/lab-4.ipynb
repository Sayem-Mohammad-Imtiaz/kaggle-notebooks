```python
data = pd.read_csv('/kaggle/input/oranges-vs-grapefruit/citrus.csv')
data.head()
```

```python
data.describe()
```

```python
final_data = pd.get_dummies(data,columns=['name'],drop_first=True)
final_data.head()
```

```python
sns.pairplot(data=final_data, hue='name_orange')
```

```python
Y = final_data['name_orange']
X = final_data.drop('name_orange', axis=1)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.5, random_state=42)
```

```python
clf = SVC().fit(X_train, Y_train)
print(classification_report(Y_test,clf.predict(X_test)))
print('conf matrix:\n',confusion_matrix(Y_test,clf.predict(X_test)))
```

```python
clf1 = SVC(kernel='linear').fit(X_train, Y_train)

print(classification_report(Y_test,clf1.predict(X_test)))
print('conf matrix:\n',confusion_matrix(Y_test,clf1.predict(X_test)))
```

```python
clf2 = SVC(kernel='sigmoid',coef0=0.1).fit(X_train, Y_train)

print(classification_report(Y_test,clf2.predict(X_test)))
print('conf matrix:\n',confusion_matrix(Y_test,clf2.predict(X_test)))
```

```python
clf2 = SVC(kernel='sigmoid',coef0=1).fit(X_train, Y_train)

print(classification_report(Y_test,clf2.predict(X_test)))
print('conf matrix:\n',confusion_matrix(Y_test,clf2.predict(X_test)))
```

```python
clf3 = SVC(kernel='poly',degree=2).fit(X_train, Y_train)

print(classification_report(Y_test,clf3.predict(X_test)))
print('conf matrix:\n',confusion_matrix(Y_test,clf3.predict(X_test)))
```

```python
clf4 = SVC(kernel='poly',degree=9).fit(X_train, Y_train)

print(classification_report(Y_test,clf4.predict(X_test)))
print('conf matrix:\n',confusion_matrix(Y_test,clf4.predict(X_test)))
```
