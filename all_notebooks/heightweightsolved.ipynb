```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/kaggle/input/heights-and-weights/data.csv')
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
print("There is {0} corelation between x & y".format(100*(np.corrcoef(df['Height'],df['Weight'])[0,1])))
```

```python
sns.scatterplot(df['Height'],df['Weight'])
```

```python
sns.distplot(df["Weight"])
```

```python
X=df['Height'].values.reshape(-1,1)
```

```python
y=df['Weight']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X,y,test_size=0.4,random_state=101)
```

```python
xtrain.shape
```

```python
ytrain.shape
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(xtrain,ytrain)
```

```python
lr.coef_
```

```python
y_pred=lr.predict(xtest)
```

```python
sns.distplot(y_pred)
```

```python
from sklearn import metrics
print("Mean Squared Error :",metrics.mean_squared_error(ytest,y_pred))
print("Root Mean Squared Error :",np.sqrt(metrics.mean_squared_error(ytest,y_pred)))
print("Mean Absolute Error :",metrics.mean_absolute_error(ytest,y_pred))
print("persentage of data points close to regression line (coff of determination) :",100*(metrics.r2_score(ytest,y_pred)))
```
