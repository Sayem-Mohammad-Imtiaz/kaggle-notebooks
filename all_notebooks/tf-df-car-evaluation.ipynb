```python
import tensorflow_decision_forests as tfdf
import pandas 
```

```python
train_df = pandas.read_csv("../input/car-evaluation/car_evaluation.csv") 
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_df, label="decision")
```

```python
# Train the model
model = tfdf.keras.RandomForestModel()
model.fit(train_ds)
```

```python
#Test the model
# Load the testing dataset
test_df = pandas.read_csv("../input/car-evaluation/car_evaluation.csv")

# Convert it to a TensorFlow dataset
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(test_df, label="decision")

# Evaluate the model
model.compile(metrics=["accuracy"])
print(model.evaluate(test_ds))


# tfdf.model_plotter.plot_model(model,tree_idx=0)
# tfdf.model_plotter.plot_model_in_colab(model, tree_idx=0)

```
