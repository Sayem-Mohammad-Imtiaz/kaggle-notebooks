```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('../input/big_data.csv', index_col = 0)
data.head()
```

```python
#Разбивка каждой строки из Model на множество, перевод в целочисленное значение и распределение по новым столбцам
data['L1'] = data['Model'].apply(lambda x: int(x.split('/')[0]))
data['L2'] = data['Model'].apply(lambda x: int(x.split('/')[1]))
data['L3'] = data['Model'].apply(lambda x: int(x.split('/')[2]))
data['Batch_Size'] = data['Model'].apply(lambda x: int(x.split('/')[3].split('B_S')[-1]))
data['Epochs'] =data['Model'].apply(lambda x: int(x.split('/')[4].split('Ep')[-1]))
```

```python
data.head()
```

```python
#Группировка. Мультииндекс. Сортировка по Score от большего.
data.groupby(['L1','L2','L3','Batch_Size','Epochs']).mean()#.sort_values(by = ['Score'], ascending = False)
```

```python
#Визуализация зависимостей
data.groupby(['L1','L2','L3','Batch_Size','Epochs']).mean().Score.unstack().plot(kind='line',figsize=(20,15), table=True, marker='o', linestyle='-')
plt.show()
```

```python
#Визуализация зависимостей
data.groupby(['L1','L2','L3','Epochs','Batch_Size']).mean().val_loss.unstack().plot(kind='line',figsize=(20,15), table=True, marker='o', linestyle='-')
plt.show()
```

**Анализ:**
1.	С увеличением количества слоёв, возрастает точность распознавания (актуально для прямого распространения). Самые высокие результаты в выборке первичного приближения, добились сети с 3 слоями.
2.	При уменьшении порога batch size (объём пакета), возрастает точность распознавания (актуально при сравнении объёмов: 250, 500, 750, 1000). Значительно увеличивается время обучения.
3.	При увеличении количества epochs обучения, возрастает точность распознавания (актуально при сравнении эпох: 20, 40, 60). Значительно увеличивается время обучения. Все нейросети, попавшие в ТОП 8, обучались при значении epochs = 60. 
4.	При визуализации данных моделей по долям ошибок, минимальное количество ошибок в моделях с batch size 1000 и epochs = 20, 40. Предположительно, получение таких значений связано с подачей batch size (объём пакета 1000, 750) при обучении нейросети. Возможно сеть «проглатывала» большой объём обучающей выборки, «размазывая» ответы, тем самым негативно влияя на точность распознавания, зато позитивно на уменьшение количества ошибок.
