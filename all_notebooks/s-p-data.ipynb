```python
import plotly.io as pio

import pandas as pd

colors = ['blue', 'orange', 'green', 'red', 'purple']

opt = []
opts = []
for i in range(0, len(colors)):
    opt = dict(
        target = df['Sector'][[i]].unique(), value = dict(marker = dict(color = colors[i]))
    )
    opts.append(opt)

data = [dict(
  type = 'scatter',
  mode = 'markers',
  x = df['Price'],
  y = df['Price/Sales'],
  text = df['Name'],
  hoverinfo = 'text',
  opacity = 0.8,
  marker = dict(
      size = df['Market Cap'],
      sizemode = 'area',
      sizeref = 200000
  ),
  transforms = [
      dict(
        type = 'filter',
        target = df['Price/Book'],
        orientation = '=',
        value = 2007
      ),
      dict(
        type = 'groupby',
        groups = df['Sector'],
        styles = opts
    )]
)]

layout = dict(
    yaxis = dict(
        type = 'log'
    )
)

fig_dict = dict(data=data, layout=layout)
pio.show(fig_dict, validate=False)
```
