## A script to find kaggle users that have reached the Number 1 rank
(Note: For some reason the Datasets section is not included in the `UserAchievements.csv` file.)

```python
import pandas as pd

UserAchievements = pd.read_csv("../input/meta-kaggle/UserAchievements.csv")

NumberOneScripts = UserAchievements.query('(HighestRanking == 1) and (AchievementType == "Scripts")')
NumberOneComp    = UserAchievements.query('(HighestRanking == 1) and (AchievementType == "Competitions")')
NumberOneDiscuss = UserAchievements.query('(HighestRanking == 1) and (AchievementType == "Discussion")')

# find the corresponding user names
Users            = pd.read_csv("../input/meta-kaggle/Users.csv")

TopScripts = pd.merge(NumberOneScripts, Users, left_on='UserId', right_on='Id')
TopComps   = pd.merge(NumberOneComp,    Users, left_on='UserId', right_on='Id')
TopDiscuss = pd.merge(NumberOneDiscuss, Users, left_on='UserId', right_on='Id')
```

## Competitions

```python
TopComps.sort_values(by='Points', ascending=False).style.bar(subset=['Points'], color='#D11313')
```

## Notebooks

```python
TopScripts.sort_values(by='Points', ascending=False).style.bar(subset=['Points'], color='#24B543')
```
