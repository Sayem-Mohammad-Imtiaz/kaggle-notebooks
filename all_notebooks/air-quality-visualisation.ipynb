```python
india=data.groupby('COUNTRY').get_group('IN')
```

```python
india
```

```python
plt.figure(figsize=(20,8))
sns.barplot(data=data,x='CITY',y='VALUE')
plt.xticks(rotation=90)
plt.show()
```

we can see that avg Air quality index of delhi,gaziabad,lucknow,muzaffarnagar,mumbai is 150+ which is unhealthy

```python
usa=data.groupby('COUNTRY').get_group('US')
```

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=usa,x='DATE',y='VALUE',color='red')
plt.show()
```

this is the air quality index  of USA variation with date

```python
plt.figure(figsize=(20,8))
sns.lineplot(data=india,x='DATE',y='VALUE',color='red')
plt.show()
```
