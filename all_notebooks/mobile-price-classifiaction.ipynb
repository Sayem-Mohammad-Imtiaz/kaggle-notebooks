```python
corr = mobile_train.corr()
plt.figure(figsize=(20,20))
sns.heatmap(corr ,annot=True,cmap="RdYlGn")
```

```python
#select featurs via corr
for index, element in corr["price_range"].items():
    if element < 0.1 and element > -0.1:
        mobile_train = mobile_train.drop([index], axis = 1)
        mobile_test = mobile_test.drop([index], axis = 1)
        
mobile_train.info()
mobile_test.info()
```

```python
y = mobile_train["price_range"]
X = mobile_train.drop("price_range", axis = 1)
```

```python
#train/test split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=44, shuffle =True)
```

```python
#k nearest 
model_knn = KNeighborsClassifier()
model_knn.fit(X_train, y_train)

#output = model_knn.predict(X_test)
print(model_knn.score(X_train, y_train))
print(model_knn.score(X_test, y_test))
```

```python
#rnd forest 
model_rndf = RandomForestClassifier()
model_rndf.fit(X_train, y_train)

print(model_rndf.score(X_train, y_train))
print(model_rndf.score(X_test, y_test))
```
