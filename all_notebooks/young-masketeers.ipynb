<a id="n1"></a>
<font color="darkblue" size=+3>Data</font>

```python
bel = pd.read_csv('/kaggle/input/beliefs-about-masks-among-young-adults/MaskBeliefs.csv', parse_dates=['Timestamp'])
```

<a id="n1"></a>
<font color="darkblue" size=+3>Profile Report</font>

```python
from pandas_profiling import ProfileReport
profile = ProfileReport(bel, title="Young Adults belief - Profiling Report")
profile.to_widgets()
profile.to_notebook_iframe()
```

<a id="n3"></a>
<font color="darkblue" size=+3>Cleaning data</font>

```python
bel['Gender'].fillna('Other',inplace=True)
bel['Gender'] = bel['Gender'].map({'Male':1,'Female':2,'Other':3})
bel['ResidentialElder'] = bel['ResidentialElder'].map({'Yes':1,'No':0})
bel['InteractedElder'] = bel['InteractedElder'].map({'Yes':1,'No':0})
bel['PreventSpread'] = bel['PreventSpread'].map({'Yes':1,'No':0})
bel['Public'] = bel['Public'].map({'Yes':1,'No':0})
bel['Boarding'] = bel['Boarding'].map({'Boarding':1,'Day':0})
bel['Reason_number'] = bel['Reason'].map({'To protect yourself AND others':1,'To protect other people':2, 'Because you are required to':3, 'To protect yourself':4, 'To protect others but also because I\'m required to':5})
bel['Restaurant'].fillna(0,inplace=True)
```

<a id="n4"></a>
<font color="darkblue" size=+3>Pre-teen (Outlier by age)</font>

```python
bel[bel['Age']<13]
```

<font color="teal" size=+2>Analysis: It is encouraging to see that a pre-teen is so responsible who believes in use, efficacy of masks and plans to wear it even when there is no compulsion.</font>

<a id="n4"></a>
<font color="darkblue" size=+3>Analysis by Reason</font>

```python
bel['Reason'].value_counts()
```

<font color="teal" size=+2>Analysis: It is heartening to note that 81% think about others or society. 14% of surveyed adults think because they need to wear a mask.</font>

<a id="n5"></a>
<font color="darkblue" size=+3>Analysis of belief by Age</font>

```python
bel.groupby(['PreventSpread','Age']).agg('PreventSpread').count()
```

<font color="teal" size=+2>Analysis: Age - wise belief analysis - 91% of 18- year olds,  93% of 17-year olds, 90% of 16-year olds, 92% of 14-years olds believe that masks do their job.
All the 15- year olds too believe in this.
</font>

<a id="n6"></a>
<font color="darkblue" size=+3>Analysis of belief by School Type</font>

```python
bel.groupby(['Boarding','PreventSpread']).agg('PreventSpread').count()
```

<font color="teal" size=+2>Analysis: 96% of school boarders are believers, while 89% of day schoolers are believers. 
</font>

<a id="n7"></a>
<font color="darkblue" size=+3>Analysis of belief by Gender</font>

```python
bel.groupby(['PreventSpread','Gender']).agg('PreventSpread').count()
```

<font color="teal" size=+2>Analysis: Gender - wise belief analysis - 96% of males believe that masks are on purpose, while 91% of females believe so. The difference is low, but the sample size is low at 107. If sample size of interviewees is say 10k, then the difference is amplified.
</font>

<a id="n7"></a>
<font color="darkblue" size=+3> When mask law is not enforced ?</font>

```python
bel.groupby(['PreventSpread','Public']).agg('PreventSpread').count()
```

<font color="teal" size=+2>Analysis: Around 93% believe that masks help in prevention of spread, but only 82% would wear masks when enforced. On the other hand, more than half of non-believers would still wear masks, even when it is not a law.
</font>

<a id="n8"></a>
<font color="darkblue" size=+3> Interaction with Vulnerable (Elders)</font>

```python
print(bel.groupby(['PreventSpread','Public','ResidentialElder']).agg('PreventSpread').count())
bel.groupby(['PreventSpread','Public','InteractedElder']).agg('PreventSpread').count()
```

<font color="teal" size=+2>Potential Risk: There is only person who does not believe in efficacy of masks, who would not wear in public (if not enforced) and who lives with a elder. So we have atleast one elderly person at risk.
There are 3 young adults who have interacted with atleast one elder.
