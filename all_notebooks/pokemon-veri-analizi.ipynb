# **NESTED FUNCTION : İÇ İÇE FONKSİYONLAR**

```python
def square():
    """return square of value"""
    def add():
        """add two local variable"""
        x = 2 
        y = 3
        z = x+y
        return z 
    return add()**2
print(square())
```

# **DEFAULT and FLEXIBLE ARGUMENTS**

```python
def f(a,b = 1, c=2):
    y = a + b + c
    return y
print(f(5))

print(f(5,4,3))
```

```python
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4)

def f (**kwargs):
    """print key and value of dictionary"""
    for key, value in kwargs.items():
        print(key, " " , value)
f(country = 'spain', capital='madrid', population= 123456)
```

# LAMBDA FUNCTION (fonksiyon yazmanı kolay ve kısa hali)

```python
#long way
def square (x):
    return x**2
print(square(5))

#short way
square = lambda x: x**2
print(square(4))
tot = lambda x,y,z: x+y+z
print(tot(1,2,3))
```
