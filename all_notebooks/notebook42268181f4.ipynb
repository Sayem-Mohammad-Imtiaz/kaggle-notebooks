```python
import tflearn
import numpy as np
from IPython.core.display import HTML, SVG, display
from random import randint

data, labels = tflearn.data_utils.load_csv('../input/training.csv', target_column=0, 
                                           categorical_labels=True, n_classes=10)
items = zip(labels, data)

# brute hack as svgwrite is not available :-)
svg = '<svg baseProfile="full" height="100px" version="1.1" width="19000px" xmlns="http://www.w3.org/2000/svg" xmlns:ev="http://www.w3.org/2001/xml-events" xmlns:xlink="http://www.w3.org/1999/xlink"><defs />'

num = {}
for ix in range(len(data)):
    y = np.argmax(labels[ix]) # convert from one-hot to number
    points = data[ix]
    num[y] = num[y] + 1 if y in num else 1
    start = [50+100 * y, 50]
    svg_path = 'M%s %s l' %tuple(start)
    for point in points:
        svg_path += ' %s' %(point)
    svg += '<path d="%s" fill="none" stroke="rgb(%s, %s, %s)" />' %(svg_path, num[y] * 5, randint(0,255), randint(0,255))

svg += '</svg>'

SVG(svg)
```
