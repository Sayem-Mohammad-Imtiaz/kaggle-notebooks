# Correlation analysis

# DATA PLOT

```python
# Line plot
# color, label, linewidth, alpha, grid, linestyle
data.Speed.plot(kind = 'line', color = 'g', label = 'Speed', linewidth = 2, alpha = 0.5, grid = True, linestyle = ':')
data.Defense.plot(color = 'r', label='Defense',linewidth=2, alpha = 0.5, grid = True, linestyle = '-.')
plot.legend(loc='upper right')
plot.xlabel('time')
plot.ylabel('Magnitude')
plot.title('Line plot')
plot.axis([0, 100, 0, 200])
plot.show()
```

```python
data1 = data.loc[:,["Attack","Defense","Speed"]]
data1.plot(subplots = True)
plot.show()
```

```python
data.plot(kind='scatter', x='Attack', y='Defense', alpha = 0.5, color='r')
plot.xlabel('Attack')
plot.ylabel('Defense')
plot.title('Attack vs Defense Scatter plot')
```

```python
data.Speed.plot(kind='hist',bins=50)
plot.show()
```
