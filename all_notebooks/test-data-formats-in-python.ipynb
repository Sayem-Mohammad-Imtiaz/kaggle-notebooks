<h1 style='background:#12AAFF; border:0; color:black'><center>npy format</center></h1> 

npy is a Python specific data format. It is a format to store numpy arrays.

```python
f = np.load("/kaggle/input/iris-dataset/iris.npy", allow_pickle=True)
print(len(f))
columns_, data_ = f
data_df = pd.DataFrame(data_, columns = columns_)
print(data_df.shape)
data_df.head()
```

<h1 style='background:#12AAFF; border:0; color:black'><center>npz format</center></h1> 

npz is a Python specific data format. It is an array to store compressed numpy arrays.

```python
f = np.load("/kaggle/input/iris-dataset/iris.npz", allow_pickle=True)
print(f.files, len(f['arr_0']))
columns_, data_ = f['arr_0']
data_df = pd.DataFrame(data_, columns = columns_)
print(data_df.shape)
data_df.head()
```

<h1 style='background:#12AAFF; border:0; color:black'><center>rds format</center></h1> 

Rds (rds) is a R-specific format for saving data.

```python
f = pyreadr.read_r('/kaggle/input/iris-dataset/iris.rds')
data_df = f[None]
print(data_df.shape)
data_df.head()
```
