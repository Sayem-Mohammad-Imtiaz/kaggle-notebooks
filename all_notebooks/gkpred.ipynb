```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import linear_model
m=pd.read_csv('../input/fifa19/data.csv')
m1=m.loc[m['Position']=='CM' ]
m2=m.loc[m['Position']=='CAM' ]
m3=m.loc[m['Position']=='CDM' ]
mf=pd.concat([m1,m2,m3])
y=mf[['Overall']]
mid=mf[['Age','Overall','Vision','BallControl','ShortPassing','Crossing','LongPassing','Interceptions','Dribbling','SprintSpeed','Balance','Stamina','Strength','LongShots','Composure','Volleys','Finishing','FKAccuracy']]
req=[]
k=mid.corr()
l=k['Overall']
for i in l.index:
    if k['Overall'][i]>0.5 and k['Overall'][i]<1:
        req.append(i)
mid=mid[req]
mid=mid.dropna()
X=mid
##lr=LinearRegression()
SGDReg =linear_model.SGDRegressor(max_iter = 1000,penalty = "elasticnet",loss = 'huber',tol = 1e-3, average = True)
Xtrain,Xtest,ytrain,ytest=train_test_split(X,y,test_size=0.33,random_state=1)
#lr.fit(Xtrain,ytrain)
SGDReg.fit(Xtrain,ytrain)
predte=SGDReg.predict(Xtest)
r2=r2_score(ytest,predte)
mse=mean_squared_error(ytest,predte)
print (r2)
print (mse)
print(SGDReg.predict(Xtest.head()))
print(ytest.head())


```
