# Categorizacion

```python
#utilizacion de encoder para asignar un valor numerico a la columna category
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()

encoder.fit(df[["category"]])

df[["category"]] = encoder.transform(df[["category"]])

print("0 = basic", "1 = luxury")
df.head()
```

# Train/Test Split

```python
#y = df["category"].copy().to_numpy().reshape(-1,1)
#X = df.drop(["category"], axis=1).copy()

X = df.drop('category', axis=1)
y = df['category']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

# Regresion logistica con scikitlearn

```python
from sklearn.linear_model import LogisticRegression

logreg_model = LogisticRegression(random_state=0, fit_intercept=False).fit(X_train, y_train)

accuracy = logreg_model.score(X_train, y_train)

test_accuracy = logreg_model.score(X_test, y_test)

print(f"accuracy: {accuracy}, test_accuracy: {test_accuracy}")
```

```python
model = logreg_model.predict(X)
model[:10]
```

```python
plt.subplots(figsize=(10,6))
plt.hist(model, bins=3)
plt.xlabel('Prediccion')
plt.ylabel('Number of records')
plt.show()
```

```python
logreg_model.predict_proba(X[:10])
```

```python
logreg_model.coef_
#Con los coeficientes se calcula las caracteristicas mas y menos relevantes
```

```python
logreg_model.intercept_
```

```python
# Log loss, aka logistic loss or cross-entropy loss

from sklearn.metrics import log_loss

log_loss(y,logreg_model.predict(X))
```

## L1

```python
c_vals = [0.05, 0.01, 0.001]

for c in c_vals:

    logreg_l1_model = LogisticRegression(C=c, penalty="l1", solver="liblinear", max_iter=1000)
    
    logreg_l1_model.fit(X_train, y_train)

    accuracy = logreg_l1_model.score(X_train, y_train)

    test_accuracy = logreg_l1_model.score(X_test, y_test)

    print(f"c: {c}, accuracy: {accuracy}, test_accuracy: {test_accuracy}")
```

```python
logreg_l1_model.coef_
```

```python
model = logreg_l1_model.predict(X_test)
model[:10]
```

```python
plt.subplots(figsize=(10,6))
plt.hist(model, bins=3)
plt.xlabel('Prediccion')
plt.ylabel('Number of records')
plt.show()
```

# L2

```python
c_vals = [100, 10, 5, 1, 0.5]

for c in c_vals:

    logreg_l2_model = LogisticRegression(C=c, penalty="l2", solver="lbfgs")
    
    logreg_l2_model.fit(X_train, y_train)

    accuracy = logreg_l2_model.score(X_train, y_train)

    test_accuracy = logreg_l2_model.score(X_test, y_test)

    print(f"c: {c}, accuracy: {accuracy}, test_accuracy: {test_accuracy}")
```

```python
logreg_l2_model.coef_
```

```python
model = logreg_l2_model.predict(X)
model[:10]
```

```python
plt.subplots(figsize=(10,6))
plt.hist(model, bins=3)
plt.xlabel('Prediccion')
plt.ylabel('Number of records')
plt.show()
```
