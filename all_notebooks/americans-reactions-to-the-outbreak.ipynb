```python
#Code from Gabriel Preda
#plt.style.use('dark_background')
def plot_count(feature, title, df, size=1):
    f, ax = plt.subplots(1,1, figsize=(4*size,4))
    total = float(len(df))
    g = sns.countplot(df[feature], order = df[feature].value_counts().index[:20], palette='RdBu')
    g.set_title("Number and percentage of {}".format(title))
    if(size > 2):
        plt.xticks(rotation=90, size=8)
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2.,
                height + 3,
                '{:1.2f}%'.format(100*height/total),
                ha="center") 
    plt.show()
```

```python
plot_count("subject", "Americans Concern about Health & Economy", df,4)
```

```python
#Code from Tanay Mehta https://www.kaggle.com/heyytanay/police-shooting-eda-interactive-map/comments

from colorama import Fore, Style

def count(string: str, color=Fore.RED):
    """
    Saves some work 
    """
    print(color+string+Style.RESET_ALL)
```

```python
def statistics(dataframe, column):
    count(f"The Average value in {column} is: {dataframe[column].mean():.2f}", Fore.RED)
    count(f"The Maximum value in {column} is: {dataframe[column].max()}", Fore.BLUE)
    count(f"The Minimum value in {column} is: {dataframe[column].min()}", Fore.RED)
    count(f"The 25th Quantile of {column} is: {dataframe[column].quantile(0.25)}", Fore.BLUE)
    count(f"The 50th Quantile of {column} is: {dataframe[column].quantile(0.50)}", Fore.RED)
    count(f"The 75th Quantile of {column} is: {dataframe[column].quantile(0.75)}", Fore.BLUE)
```

```python
# Print Offset Column Statistics
statistics(df, 'samplesize')
```
