# **Explore tha data**

# **Data cleaning and wrangling**
* Drop "Agency", "Neighborhood", "sq_mtrs", "link" columns and all rows with zero/NaN values
* Convert Price" to ints
* Convert "Neighborhood" column to ints

```python
# Drop "Neighborhood", "sq_mtrs", "link" columns and all rows with zero/NaN values
del data["Agency"]
del data["Neighborhood"]
del data["sq_mtrs"]
del data["link"] 
data = data.dropna()
data
```

```python
# Convert Price" to ints
data["Price"] = data["Price"].str.replace("KSh ", "")
data
data["Price"] = data["Price"].str.replace(",", "")
data["Price"] = pd.to_numeric(data["Price"])
data.info()
```

```python
data.corr()
```

# **Set features and labels**

```python
X = data[["Bedrooms", "Bathrooms"]]
y = data[["Price"]]
```

# **Train,test and fit model**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.40, random_state = 32)
```

```python
clf = DecisionTreeRegressor()
clf.fit(X_train, y_train)
```

# **Prediction and accuracy**

```python
clf.predict(X_test)[:5]
```

```python
str(y_test[:5])
```

```python
# Accuracy
clf.score(X_test, y_test)
```
