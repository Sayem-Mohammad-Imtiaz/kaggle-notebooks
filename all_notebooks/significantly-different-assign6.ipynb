**Required APIs**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier 
from sklearn.metrics import classification_report, confusion_matrix  
from sklearn.cluster import KMeans
from scipy.cluster.vq import whiten
```

**KNN Classificaation**

```python
data = pd.read_csv("../input/Absenteeism_at_work.csv")
#1st column to 13th column
X =  data.iloc[:,range(0,13)]
#14st column --> Labels
Y =  data.iloc[:,14]

#normalising the data
normal_X = preprocessing.normalize(X)
```

```python
#Splitting of data set
Train_X, Test_X, Train_y, Test_y = train_test_split(X, Y, test_size = 0.2, random_state = 239)
Train_y = Train_y.ravel()
Test_y = Test_y.ravel()
```

```python
k = []
accuracy = []
for i in range(int(np.sqrt(data.shape[0]))):
    K_value = i+1
    neigh = KNeighborsClassifier(n_neighbors = K_value, weights='uniform', algorithm='auto')
    neigh.fit(Train_X, Train_y) 
    Pred_y = neigh.predict(Test_X)
    #print(Pred_y)
    k.append(K_value)
    accuracy.append(accuracy_score(Test_y,Pred_y)*100)
    
```

```python
print("Accuracy of the model: ",max(accuracy))
print("K value: ",accuracy.index(max(accuracy))+1)
```

```python
#graph for visualization
plt.plot(k, accuracy)
plt.xlabel('K Values')
plt.ylabel('Accuracy')
plt.title('K values vs Accuracy graph')
plt.show()
```

**Decision Tree**

Creteria -->GINI

```python
clf_gini = DecisionTreeClassifier(criterion = "gini", random_state = 100, max_depth=3, min_samples_leaf=5)
clf_gini.fit(Train_X, Train_y)
Pred_y_gini = clf_gini.predict(Test_X)
```

```python
print(accuracy_score(Test_y,Pred_y_gini)*100)
```

```python
print(confusion_matrix(Test_y,Pred_y_gini))
```

```python
print(classification_report(Test_y,Pred_y_gini))
```

**Decision Tree**

Creteria --> Entropy

```python
clf_entropy = DecisionTreeClassifier(criterion = "entropy", random_state = 100, max_depth=3, min_samples_leaf=5)
clf_entropy.fit(Train_X, Train_y)
Pred_y_entropy = clf_entropy.predict(Test_X)
```

```python
print(accuracy_score(Test_y,Pred_y_entropy)*100)
```

```python
print(confusion_matrix(Test_y,Pred_y_entropy))
```

```python
print(classification_report(Test_y,Pred_y_entropy))
```

**Why the value of k is chosen to be that? **

We know that the k-value should be an odd number because if it is even there are chances of ties and ambiguity as to which class to be placed in. Also k value shouldn’t be a multiple of the number of classes. As a rule of thumb we know that the value of k is usually taken as square root of n.  So here we checked the accuracy value for every k in the range of [1, square root of n]. 
We found that for k = 7, we have the highest accuracy of 49.32% for the given split of train and test data.

**Comparison of models**

The accuracies we got for both the techniques, that is kNN and decision trees is pretty much the same.
The accuracy we got for KNN is 49.32%.
For decision trees , we got an accuracy of 51.35%. ( this is when we built the decision tree based on the gini index).
KNN is generally good for continuous value inputs as it is in this case.
It is also  good for instances where there are less target variables and more number of data points. But here the number of target variables is high (i.e; 15) . So maybe , that’s the reason we have a slightly higher accuracy in case of decision trees than in case of KNN.
