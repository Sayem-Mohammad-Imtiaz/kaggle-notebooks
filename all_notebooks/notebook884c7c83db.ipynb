```python
import os

import numpy as np

import pandas as pd

import numpy as np, pandas as pd

import matplotlib.pyplot as plt

from sklearn import metrics , model_selection

## Import the Classifier.

from sklearn.naive_bayes import GaussianNB
```

**LOAD THE DATA SET**

```python
data = pd.read_csv('/kaggle/input/car-evaluation-data-set/car_evaluation.csv',names=['buying','maint','doors','persons','lug_boot','safety','class'])
data.head()
```

**CHECK SOME INFORMATION ABOUT DATA SET**

```python
data.info()
```

**IDENTIFY THE TARGET VARIABLE**

```python
data['class'],class_names = pd.factorize(data['class'])
```

```python
print(class_names)

print(data['class'].unique())
```

***Identify the predictor variables and encode any string variables to equivalent integer codes ***

```python
data['buying'],_ = pd.factorize(data['buying'])

data['maint'],_ = pd.factorize(data['maint'])

data['doors'],_ = pd.factorize(data['doors'])

data['persons'],_ = pd.factorize(data['persons'])

data['lug_boot'],_ = pd.factorize(data['lug_boot'])

data['safety'],_ = pd.factorize(data['safety'])

data.head()
```

***Check the data types now ***

```python
data.info()
```

**Select the predictor feature and select the target variable**

```python
X = data.iloc[:,:-1]

y = data.iloc[:,-1]
```

**Train test splits**

```python
# split data randomly into 70% training and 30% test

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.3, random_state=123)
```

***Training / model fitting ***

```python
model = GaussianNB()

## Fit the model on the training data.

model.fit(X_train, y_train)
```

***Model parameters study ***

```python
# use the model to make predictions with the test data

y_pred = model.predict(X_test)

# how did our model perform?

count_misclassified = (y_test != y_pred).sum()

print('Misclassified samples: {}'.format(count_misclassified))

accuracy = metrics.accuracy_score(y_test, y_pred)

print('Accuracy: {:.2f}'.format(accuracy))
```
