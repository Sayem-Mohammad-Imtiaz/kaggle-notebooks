```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import matplotlib.colors as mcolors
```

```python
df = pd.read_csv('../input/windows-store/msft.csv')
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
sns.set(rc={'figure.figsize':(10,5)})
ax = sns.countplot(x="Category", data=df)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

The most frequently appearing category in this dataset is Music.

```python

num_of_rows = len(df.index)

free_apps=float((df["Price"] == 'Free').sum() / num_of_rows * 100)
paid_apps =float((df["Price"] != "Free").sum() / num_of_rows*100)


y = [free_apps,paid_apps]
x = ["Free Apps","Paid Apps"]

```

```python
plt.pie(y, labels=x, autopct='%1.0f%%', pctdistance=0.8, labeldistance=1.2)
```

97% of the apps featured in this dataset are available to download for free.

```python
df1 = df.sort_values(by = 'Rating', ascending=False).head(100)
sns.set(rc={'figure.figsize':(10,5)})
ax = sns.countplot(x="Category", data=df1)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```
