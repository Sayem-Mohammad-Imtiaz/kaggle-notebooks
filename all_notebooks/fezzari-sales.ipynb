<h2>DATA EXPLORATION</h2>

<h3>1) What was the best month for sales? How much was earned that month?</h3>

<h4>Calculating Sales Cost</h4>

```python
df['Sales Cost'] = df['Quantity']* df['Unit Cost']
df
```

<h4>Grouping Month by Sales Cost</h4>

```python
df_temp = df.groupby('Month').sum()['Sales Cost'].reset_index()
df_temp
```

<h4>Plotting Sales by Month</h4>

```python
import matplotlib.pyplot as plt

months = range(1,13)
print(months)

plt.bar(months,df.groupby(['Month']).sum()['Sales Cost'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```

Result:<h4>Here you can see that Month 6 has high Sales.So June is the best month for sales with $2130992.54</h4>

<h3>2) Which Country is more profitable?</h3>

<h4>Grouping Revenue by Country</h4>

```python
new_df=df.groupby(['Country']).sum()
new_df.loc[:,'Revenue']
```

<h4>Plotting Revenue by Country</h4>

```python
keys = [Country for Country, df in df.groupby(['Country'])]
plt.bar(keys,df.groupby(['Country']).sum()['Revenue'])

plt.ylabel('Sales in USD ($)')
plt.xlabel('Country')
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

Result:<h4>Here you can see that United States has sold most products with $10371147.0 revenue</h4>

<h3>3)Which product category and subcategory are more profitable?</h3>

<h4>Sorting Product Category and Subcategory based on revenue</h4>

```python
cat_subcat = pd.DataFrame(df.groupby(['Product Category', 'Sub Category']).sum()['Revenue'])
cat_subcat.sort_values(['Product Category','Revenue'], ascending=False)
```

Result:<h4>Here you can see<br>i)In Clothing,jerseys are more profitable<br>ii)In Bikes,Mountain Bikes are more profitable<br>iii)In Accessories Tires and Tubes are more profitable </h4>

<h3>4)What are the top 10 product by Sales?</h3>

<h4>Sorting top 10 Products by Sales Cost</h4>

```python
product_sales = pd.DataFrame(df.groupby('Sub Category').sum()['Sales Cost'])
product_sales.sort_values(by=['Sales Cost'], inplace=True, ascending=False)
product_sales.head(10)
```

<h4>Plotting top 10 Products by Sales Cost</h4>

```python
keys = [SubCategory for SubCategory, df in df.groupby(['Sub Category'])]
plt.bar(keys,df.groupby(['Sub Category']).sum()['Sales Cost'])

plt.ylabel('Sales in USD ($)')
plt.xlabel('Sub Category')
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```

Result:<h4>Here you can see top 10 products by sales.And Mountain Bikes holds first in sales with $5029503.58 sales cost.</h4>

<h3>5)Which is the most sold product?</h3>

<h4>Grouping and Sorting products based on Quantity</h4>

```python
most_selling_products = pd.DataFrame(df.groupby('Sub Category').sum()['Quantity'])
most_selling_products.sort_values(by=['Quantity'], inplace=True, ascending=False)
most_selling_products[:10]
```

<h4>Plotting most sold products based on Quantity</h4>

```python
product_group = df.groupby('Sub Category')
quantity = product_group.sum()['Quantity']

keys = [pair for pair, df in product_group]
plt.bar(keys, quantity)
plt.xticks(keys, rotation='vertical', size=8)
plt.ylabel('Quantity')
plt.xlabel('Sub Category')
plt.show()
```

```python
category_chart=df['Sub Category'].value_counts()
category_chart.plot(kind = 'pie', autopct='%1.1f%%', figsize=(15, 15)).legend()
```

Result:<h4>Here you can see Tires and Tubes have sold more with 22201 sales quantity.</h4>
