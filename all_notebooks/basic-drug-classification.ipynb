```python
df_upd=pd.get_dummies(df,columns=['Sex','BP','Cholesterol'])
```

```python
df_upd
```

```python
def multi(name):
    Drug_cat=1
    if name=='DrugY':
        Drug_cat=0
    elif name=='drugC':
        Drug_cat=1
    elif name=='drugX':
        Drug_cat=2

    elif name == 'drugB':
        Drug_cat=3
    else:
        Drug_cat=4
    return Drug_cat
    
```

```python
df_upd['Drug_cat']=df_upd['Drug'].apply(multi)
```

```python
df_upd
```

```python
df_model=df_upd.drop('Drug',axis=1)
```

```python
X=df_model.drop('Drug_cat',axis=1)
```

```python
y=df_model['Drug_cat']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(X, y,test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=10)
```

```python
model.fit(X_train,y_train)
```

```python
pred_tr=model.predict(X_train)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(pred_tr,y_train))
```

```python
pred_te=model.predict(X_test)
```

```python
print(classification_report(pred_te,y_test))
```
