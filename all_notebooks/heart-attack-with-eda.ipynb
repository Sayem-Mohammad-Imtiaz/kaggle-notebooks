### EDA

#### Statistical Summary

```python
df_stat = df[['age', 'cp', 'trtbps', 'chol', 'thalachh', 'oldpeak', 'slp','caa', 'thall']].copy()
df_stat.describe()
```

We have seen the statistical summary for the data, Now let's move forward to visualize the insights of the data.  

#### Count of 'Chance of Heart attack'

```python
plt.figure(figsize=(14,6))
df_output = df.copy()
df_output['output'] = df_output['output'].map({0: "Lower chance of Heart Attack", 1: 'Higher chance of Heart Attack'})
sns.countplot(data= df_output, x = 'output', palette= 'dark')
```

By observing the above data, we came to know that data has been fairly distributed and count of 'Higher chance of Heart attack' is high seen from the dataset. 

#### Distribution of Age and the Chance of Heart Attack

```python
plt.figure(figsize=(16,10), dpi = 250)
df_age = df[['age', 'output']].copy()
df_age.columns = ['Age', 'Chance of Heart Attack']
df_age['Chance of Heart Attack'] = df_age['Chance of Heart Attack'].map({0: 'Lower Chance of Heart Attack', 1: 'Higher chance of Heart Attack'})
sns.countplot(data = df_age, x = 'Age', hue = 'Chance of Heart Attack', palette='dark')
```

```python
sns.jointplot(data = df_age, x = 'Chance of Heart Attack', y = 'Age', height=10, s = 100, alpha = 1)
```

```python
df_age_corr = df[['age', 'output']].copy()
corr_age= df_age_corr.corr()
plt.figure(figsize= (15,7))
sns.heatmap(corr_age, annot= True, linewidths= 2)
plt.title('Correlation Matrix for the distribution of Age and the Chance of the Heart Attack', fontsize= 10)
plt.show()
```

We clearly see that, from the first graph, there is the higher chance of heart attack from the age between 41 and 45 years, then after that,between 51 and 54 years, there is higher values. So, age is not properly correlate with the chance of the heart attack. From the second graph, it representing the same as the first graph. From the third, we clearly see that, there is no correlation between age and chance of heart attack.

#### Distribution of Sex and the Chest Pain

```python
df_sex_cp = df[['sex','cp']]
df_sex_cp.columns = ['Sex', 'Type of Chest Pain']
df_sex_cp['Sex'] = df_sex_cp['Sex'].map({0: 'Women', 1: 'Men'})
df_sex_cp['Type of Chest Pain'] = df_sex_cp['Type of Chest Pain'].map({0: 'Typical Angina',
                                                                       1: 'Atypical Angina',
                                                                       2: 'Non-Anginal Pain',
                                                                       3: 'Asymptomatic'})
plt.figure(figsize=(14,7), dpi = 200)
sns.countplot(data=df_sex_cp, x= 'Type of Chest Pain', hue= 'Sex', palette = 'icefire', order=['Typical Angina', 'Atypical Angina','Non-Anginal Pain','Asymptomatic'])
```

```python
df_cp_corr = df[['sex', 'cp']].copy()
df_cp_corr.columns = ['Sex', 'Chest Pain']
corr_cp = df_cp_corr.corr()
plt.figure(figsize= (14,7), dpi = 90)
sns.heatmap(corr_cp, annot= True, linewidths=2)
plt.title('Correlation Matrix', fontsize = 10)
plt.show()
```

#### Distribution of Age and the Chest Pain

```python
df_age_cp = df[['age', 'sex', 'cp']].copy()
df_age_cp.columns = ['Age', 'Sex', 'Type of Chest Pain']
df_age_cp['Sex'] = df_age_cp['Sex'].map({0: 'Women', 1: 'Men'})
df_age_cp['Type of Chest Pain'] = df_age_cp['Type of Chest Pain'].map({0: 'Typical Angina',1: 'Atypical Angina',2: 'Non-Anginal Pain',3: 'Asymptomatic'})
plt.figure(figsize=(12,7), dpi = 100)
sns.violinplot(data=df_age_cp, x = 'Type of Chest Pain', y = 'Age', hue = 'Sex', palette = 'dark',order=['Typical Angina', 'Atypical Angina','Non-Anginal Pain','Asymptomatic'])
```

```python
df_age_cp_corr = df[['sex', 'cp', 'output']].copy()
df_age_cp_corr.columns = ['Sex', 'Chest Pain', 'Chance of Heart Attack']
corr_age_cp = df_age_cp_corr.corr()
plt.figure(figsize= (14,7), dpi = 90)
sns.heatmap(corr_age_cp, annot= True, linewidths=2)
plt.title('Correlation Matrix', fontsize = 10)
plt.show()
```
