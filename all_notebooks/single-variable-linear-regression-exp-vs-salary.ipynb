#### Now, we need to find the equation for above visualized line, which can be found by using sklearn's linear_model 

```python
from sklearn import linear_model

linear_reg = linear_model.LinearRegression()
```

```python
# for x dataframe, we will drop salary as salary is our y
df_x = df.drop(['Salary'], axis='columns')
df_x
```

```python
# now we will make our y ready
y = df.Salary
y
```

```python
# Now we will train the model here
reg = linear_reg.fit(df_x,y)
```

```python
# let's check the prediction with case : Experience is 5.5 years
reg.predict([[5.5]])
```

```python
# let's check the coeffiecient and intercept 
reg.coef_
```

```python
reg.intercept_
```

```python
# So if we use the formula here, which is y = coefficient(x) + intercept, where coef, x and intercept are (9449.96232146, 5.5, 25792.20019866871)
y_prediction = 9449.96232146*5.5 + 25792.20019866871
y_prediction
```

```python
# Now we will split our data in train test data sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df_x,y,test_size=0.2)
```

```python
# lets check the rows in X_test and X_train
len(X_test)
```

```python
X_test
```

```python
len(X_train)
```

```python
# Now will train in the train datasets with our linear regression model
reg_train = linear_reg.fit(X_train, y_train)
```

```python
# Now we will predict the X_test
reg_test_predict = reg_train.predict(X_test)
reg_test_predict
```

```python
# Now we can check the score
reg_score = reg_train.score(X_test, y_test)
reg_score
```
