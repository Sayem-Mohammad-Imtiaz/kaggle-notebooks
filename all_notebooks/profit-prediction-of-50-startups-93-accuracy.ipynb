```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
data = pd.read_csv('../input/50-startups/50_Startups.csv')
data.head()
```

```python
X=data.iloc[:,:4]
y=data.iloc[:,4]
```

```python
X.head()
```

```python
y.head()
```

**Since 'state' has categorical data, so converting the data into dummy variables**

```python
states=pd.get_dummies(data['State'],drop_first=True)
states.head()
```

```python
X_data=pd.concat([X,states],axis=1).drop('State',axis=1)
X_data.head()
```

# Scaling The Data

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X_data)
X_scale=pd.DataFrame(X_scaled)
X_scale.columns=['R&D Spend','Administration','Marketing Spend','Florida','New York']
X_scale.head()
```

# Splitting The Data

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scale,y,test_size=0.2,random_state=0)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

# Creating The Model

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
y_pred
```

# Accuracy of the Data

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```
