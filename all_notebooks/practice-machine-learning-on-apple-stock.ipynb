# Machine learning practice project
### This code below is trying to predict the Apple stock using XGBoost

```python
import pandas as pd
data = pd.read_csv("../input/stock-market-data/stock_market_data/nasdaq/csv/AAPL.csv")
data.tail()
```

### Clean the dataset

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer()
data = data[['Low', 'Open', 'Volume', 'High', 'Close', 'Adjusted Close']]
imputed_data = pd.DataFrame(imputer.fit_transform(data))
imputed_data.columns = data.columns
data = imputed_data
```

### Load the train and validation data

```python
from sklearn.model_selection import train_test_split
X = data[['Low', 'Open', 'Volume', 'Close', 'Adjusted Close']]
y = data.High

X_train, X_valid, y_train, y_valid = train_test_split(X, y)
```

### Begin to create the model and predict the outcome

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error

model = XGBRegressor(n_estimators=1000, learning_rate=0.05)
model.fit(X_train, y_train,
              early_stopping_rounds=5, 
              eval_set=[(X_valid, y_valid)],
              verbose=False)

pred = model.predict(X_valid)
print("Mean Absolute Error: " + str(mean_absolute_error(pred, y_valid)))
```

### Compare the prediction with the validation data

```python
real = []
for item in y_valid.iteritems():
    real.append(item[1])
compare = pd.DataFrame(data={
    'Predictions':pred,
    'Real':real
})

compare
```
