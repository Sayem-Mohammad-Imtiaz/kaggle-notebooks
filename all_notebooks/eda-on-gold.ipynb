## So there are only two columns.

## so lets make another 2 columns which contains year and month
  - Then we can easily do some analysis on this data set

```python
df["year"]=pd.DatetimeIndex(df["Date"]).year
```

```python
df["month"]=pd.DatetimeIndex(df["Date"]).month
```

```python
df.head()
```

## so we make another two columns based on that we can do some EDA.

# 01. i) On which year the Gold price was highest

```python
df.loc[df["Price"]==df["Price"].max()]
```

# In year 2020 & month 7th. The gold price was highest.

## ii) On which year the Gold price was lowest

```python
df.loc[df["Price"]==df["Price"].min()]
```

# In the year 1952 & 1st,2nd month the price was lowest

## 02. Now lets see the year wise Trend

```python
x=df.year
y=df.Price

 
sns.set(style='whitegrid')

sns.scatterplot(x,y)
plt.show()
```
