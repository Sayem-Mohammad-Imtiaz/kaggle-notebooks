```python
train = pd.read_csv("/kaggle/input/music-genre-classification/train.csv")
train = train.fillna(0)
#print(train.head(5))


test = pd.read_csv("/kaggle/input/music-genre-classification/test.csv")
test = test.fillna(0)
print(test.head(5))
```

```python
df_test = test
del df_test["Artist Name"]
del df_test["Track Name"]
print(df_test.head(2))
```

```python
df_train = train
genre = df_train.pop("Class") 
del df_train["Artist Name"]
del df_train["Track Name"]
print(df_train.head(5))
#slk.ensemble.RandomForest()
```

```python
model = sklearn.ensemble.RandomForestClassifier()
model.fit(df_train, genre)

```

```python
pred = model.predict(test)
pred = pred.ravel()
print(pred)
```

```python
pred_one_hot = np.eye(11)[pred]
print(pred_one_hot)
type(pred_one_hot)
```

```python
result = pd.DataFrame(pred_one_hot)
result.to_csv('mysubmissionfile.csv', index=False)
```
