Input data

```python
import pandas as pd
appstore_games = pd.read_csv("../input/17k-apple-app-store-strategy-games/appstore_games.csv")
appstore_games
```

Use a histogram to see the popularity of the strategy games. To describe the popularity, we can use "average rating".

```python
sns.distplot(a=appstore_games['Average User Rating'].dropna(), kde=False)
```

As we can see, nearly 3000 games receive a 4.5 rating, the second highest rating. More than half of the game receive a rating above 4.0. This may suggest the strategy games are favored.

We can also create a KDE plot to deduce the distribution of "average user rating".

```python
sns.kdeplot(data=appstore_games['Average User Rating'].dropna(), shade=True)
```

However, the games which receive the high score may not draw the most attention. If we take "user rating count" into account, we may see a different conclusion. Therefore, we should consider the 2D KDE plot to test whether a high score draws the most attention.

```python
sns.jointplot(x=appstore_games['Average User Rating'].dropna(), y=appstore_games['User Rating Count'].dropna(), kind="kde",ylim=300000)
```

As we can see, most games only receive less than 10k ratings. Although the average rating is high, most games fail to draw people's attention. Notice that there is a dark area in the middle. This means some games attract 150k ratings, and the rating is as high as 4.5. From this dataset, it seems that only a small part of strategy games are very popular. However, "average user rating" and "user rating" are not the only indicator of the popularity. For example, many people would not rate a game even if they like it. In addition, the dataset is too small because there are only 17000 rows, and many ones are dropped. A small dataset cannot describe a pattern.
