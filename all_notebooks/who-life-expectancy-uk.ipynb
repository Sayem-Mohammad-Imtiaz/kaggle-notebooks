Import

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

Read

```python
df = pd.read_csv('/kaggle/input/life-expectancy-who/Life Expectancy Data.csv')
df
```

Convert year to year

```python
import datetime

df.Year = pd.to_datetime(df.Year,format = '%Y')
df.Year 
```

Analyse

```python
df.info()
```

Country

```python

country = df['Country'].to_list()

# using naive method to remove duplicates from list  
list_country = [] 
for i in country: 
    if i not in list_country: 
        list_country.append(i) 

list_country
```

```python
print(len(list_country))
```

Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

corrmat = df.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True)
```

Countries

```python

country = "United Kingdom of Great Britain and Northern Ireland"

include_uk = df[df['Country'].values == country]
exclude_uk = df[df['Country'].values != country]
uk_life = include_uk
uk_life
```
