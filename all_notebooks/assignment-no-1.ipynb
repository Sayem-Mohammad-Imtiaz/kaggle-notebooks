### Creating Binning Automatically using linspace function

```python
df1['Killed'].plot(kind='hist')
```

```python
bins = np.linspace(min(df1['Killed']), max(df1['Killed']),4)
bins
```

```python
Group_name=['Low','Medium','High']
df1['Killed-bin']=pd.cut(df1['Killed'], bins, labels=Group_name, include_lowest=True)
```
