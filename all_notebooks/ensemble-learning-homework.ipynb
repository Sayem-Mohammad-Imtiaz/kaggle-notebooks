Read the training dataset

```python
file_path = '../input/mnist-in-csv/mnist_train.csv'
missing_value_formats = ["n.a.","?","NA","n/a", "na", "--"]

train_dataset = pd.read_csv(file_path, na_values = missing_value_formats)
print(train_dataset.head(10))
```

Explore the dataset

# Prepare for training

Define X and y

```python
X = train_dataset.iloc[:, 1:]
print(X)
y = train_dataset.iloc[:, 0]
print(y)
```

Split the dataset in 2 parts:
* training
* validation

```python
train_X, X_validation, train_y, y_validation = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

# 1. Voting Classifier with RandomForrestClassifier and DecisionTreeClassifier

```python
model1 = RandomForestClassifier(random_state=1)
model2 = DecisionTreeClassifier(random_state=1)
model = VotingClassifier(estimators=[('lr', model1), ('dt', model2)], voting='hard')
start_time = time.time()
model.fit(train_X, train_y)
print("The Voting Classifier took: %s seconds." % (time.time() - start_time))
predictions = model.predict(X_validation)
print(predictions)
#model.score(X_validation,y_validation)
```

```python
print("The accuracy for the Voting Classifier model is:", accuracy_score(y_validation, predictions) *100, '%')
```

# 2. BaggingClassifier with RandomForestClassifier

```python
model = BaggingClassifier(RandomForestClassifier(random_state=1), 3)
start_time = time.time()
model.fit(train_X, train_y)
print("The BaggingClassifier took: %s seconds." % (time.time() - start_time))
model.fit(train_X, train_y)
predictions = model.predict(X_validation)
```

```python
print("The accuracy for the BaggingClassifier model is:", accuracy_score(y_validation, predictions) *100, '%')
```

# 3. XGBoost

```python
model=xgb.XGBClassifier(random_state=1,learning_rate=0.01)
start_time = time.time()
model.fit(train_X, train_y)
print("The XGBoost took: %s seconds." % (time.time() - start_time))
print("The accuracy for the XGBoost model is:", model.score(X_validation,y_validation)*100, '%')
```
