```python
import csv
csv_filename = "../input/clash-royale-season-18-dec-0320-dataset/BattlesStaging_01042021_WL_tagged/BattlesStaging_01042021_WL_tagged.csv"
chunk_size = 500000

text_file_reader = pd.read_csv(csv_filename, engine='python',encoding='utf-8-sig', quoting=csv.QUOTE_MINIMAL, chunksize = chunk_size, index_col=0)

print('done')
```

```python
dfList = []
counter = 0

for df in text_file_reader:
    dfList.append(df)
    counter= counter +1
    print("Max rows read: " + str(chunk_size * counter) )

print('done')
```

```python
df = pd.concat(dfList,sort=False)
print('concatenation complete')
```

```python
df.shape
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter

ax = df.hist('average.startingTrophies', bins = 25, grid = False, figsize = (28, 16), color = '#86bf91', zorder = 2, rwidth = 0.9)

ax = ax[0]
for x in ax:

    x.spines['right'].set_visible(False)
    x.spines['top'].set_visible(False)
    x.spines['left'].set_visible(False)

    x.tick_params(axis = 'both', which = 'both', bottom = 'off', top = 'off', labelbottom = 'on', left = 'off', right = 'off', labelleft = 'on')

    vals = x.get_yticks()
    for tick in vals:
        x.axhline(y = tick, linestyle = 'dashed', alpha = 0.4, color = '#eeeeee', zorder = 1)

    x.set_title('Trophy Distribution', pad = 30, weight = 'bold', size = 24)

    x.set_xlabel('Average Trophies', labelpad = 20, weight = 'bold', size = 12)

    x.set_ylabel('Matches', labelpad = 20, weight = 'bold', size = 12)

    x.yaxis.set_major_formatter(StrMethodFormatter('{x:,g}'))
```

```python
x.figure.savefig('trophy_distribution.png')

print('save complete')
```

```python
def level_disparity(row):

    return row['winner.totalcard.level'] - row['loser.totalcard.level']

df['level.disparity'] = None
df['level.disparity'] = df.apply(level_disparity, axis = 1)

print('level disparities calculated')
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter

bx = df.plot.scatter('average.startingTrophies', 'level.disparity', figsize = (28, 16), color = '#86bf91')
```

```python
bx.figure.savefig('level_disparity.png')

print('save complete')
```

```python
final_column_list = [ 
 'average.startingTrophies',
 'winner.startingTrophies',
 'winner.trophyChange',
 'winner.crowns',
 'loser.startingTrophies',
 'loser.trophyChange',
 'loser.crowns',
 'winner.totalcard.level',
 'winner.elixir.average',
 'loser.totalcard.level',
 'loser.elixir.average']

print('final column list initialized')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

f, ax = plt.subplots(figsize = (16, 12))
corr = df[final_column_list].corr()
sns_plot = sns.heatmap(corr, mask = np.zeros_like(corr, dtype = np.bool_), cmap =sns.diverging_palette(220, 10, as_cmap = True), square = True, ax = ax)
```

```python
sns_plot.figure.savefig('correlation_matrix.png')

print('save complete')
```
