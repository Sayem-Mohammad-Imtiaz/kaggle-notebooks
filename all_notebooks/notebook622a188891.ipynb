```python
import statsmodels.api as sm
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=1,test_size=0.2)
```

```python
print("X_train : ",X_train.shape)
print("y_train : ",X_train.shape)
print("X_test : ",X_test.shape)
print("y_test : ",X_test.shape)
```

```python
linreg = sm.OLS(y_train,X_train).fit()
print(linreg.summary())
```

```python
y_pred = linreg.predict(X_test)
actual_claim = y_test
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from statsmodels.tools.eval_measures import rmse
```

```python
linreg_rmse = rmse(actual_claim, y_pred)
linreg_rsquared = linreg.rsquared
linreg_rsquared_adj = linreg.rsquared_adj 
```

```python
score_card = pd.DataFrame(columns=['R-Squared', 'Adj. R-Squared', 'RMSE'])
score_card
```

```python
model_metrics = pd.Series({'RMSE':linreg_rmse,'R-Squared': linreg_rsquared,'Adj. R-Squared': linreg_rsquared_adj })
score_card = score_card.append(model_metrics, ignore_index=True)
score_card
```
