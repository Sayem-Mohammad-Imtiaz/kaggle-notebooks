```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
data = pd.read_csv("/kaggle/input/heart-disease-uci/heart.csv")
```

```python
data.info()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
data[["age","trestbps","chol","thalach","oldpeak"]] = sc.fit_transform(data[["age","trestbps","chol","thalach","oldpeak"]])
```

```python
data
```

```python
data[["cp","fbs","restecg","exang","slope","ca","thal"]] = data[["cp","fbs","restecg","exang","slope","ca","thal"]].astype(object)
```

```python
data=pd.get_dummies(data)
```

```python
X = data.copy()
y = data["target"]
```

```python
X.drop(columns="target",axis=1,inplace=True)
```

```python
X = X.values
y=y.values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.18,random_state=99)
```

```python
X_train.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
ac=accuracy_score(y_test, pred)
ac
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
pred1 = dt.predict(X_test)
```

```python
ac1=accuracy_score(y_test, pred1)
ac1
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train,y_train)
```

```python
pred4 = nb.predict(X_test)
```

```python
ac4=accuracy_score(y_test, pred4)
ac4
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=99)
lr.fit(X_train, y_train)
```

```python
pred5 = lr.predict(X_test)
```

```python
ac5=accuracy_score(y_test, pred5)
ac5
```
