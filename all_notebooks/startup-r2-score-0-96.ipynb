```python
from sklearn.metrics import mean_absolute_error,r2_score,mean_squared_error
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(data[['R&D Spend','Administration']],data['Profit'],test_size=0.2,random_state=101)
```

```python
lr=LinearRegression()
lr.fit(train_x,train_y)
predict=lr.predict(test_x)
print('MAE',mean_absolute_error(predict,test_y))
print('MSE',mean_squared_error(predict,test_y))
print('RMSE',np.sqrt(mean_squared_error(predict,test_y)))
print('R2',r2_score(predict,test_y))
```
