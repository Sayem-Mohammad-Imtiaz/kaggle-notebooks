## How movie genre populiarity changed in 2006 - 2016

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Importing data
# I was using the data from Kaggle 
# There's the link to my data:https://www.kaggle.com/PromptCloudHQ/imdb-data
filmai = pd.read_csv("../input/imdb-data/IMDB-Movie-Data.csv", index_col=0)
```

```python
# selecting the columns that I am going to use for my analysis
df=filmai[['Genre', 'Year']].copy()
```

```python
#Spliting Genre column, so I could see wich spesific genre was repeated the most
df.Genre.apply(lambda x: x.split(",") )
```

```python
# I create a list with all genre's from Genre column
genre_data = list()
for idx in df.index:
    genre_data+=df.loc[idx, 'Genre']. split(",")  
```

```python
#I convert my list to set for having only uniqe values. 
genre_list=set(genre_data)
```

```python
# I create a new list using for loop, to check how many times uniqe genre was repaeted 
# and then convert it to a dataframe
total= []
for genre in genre_list:
    _=filmai.Genre.apply(lambda x: genre in x)
    total.append([genre, _.sum()])
    #print(genre, _.sum())
total = pd.DataFrame(total,columns = ['Genre','Sum'])
total = total.sort_values(by='Sum',ascending=False)
total
```

```python
#visualing data using bar chart
plt.figure(figsize=(10,6))
plt.title("Genre populiarity acording to IMDB 1000 popular movies off 2006-2016")
sns.barplot(y=total.Genre, x=total.Sum)
```

```python
# I created a new dataframe based on my previos one and added the most popular genre's as new colums to the new dataframe
# and I used Boolean Indexing to check if the genre was in all the movies and converted Boolean to integer indexing. 
nf1 = df.copy()
genre_list_2 = ['Action', 'Drama', 'Horror','Thriller','Comedy','Romance','Adventure','Crime']

for genre in genre_list_2:
    nf1[genre] = nf1.Genre.apply(lambda x: genre in x).astype(int)

nf1
```

```python
# I modify my dataframa with only the column's that I will use further and removed all NaN values
nf1=nf1[['Year',"Action","Drama","Horror","Thriller","Comedy","Romance","Adventure","Crime"]].dropna()
nf1
```

```python
# Created 9 new dataframes for individual genre to check how many times they appear every year
action = nf1.groupby(['Year']).Action.agg([sum])
drama = nf1.groupby(['Year']).Drama.agg([sum])
horror = nf1.groupby(['Year']).Horror.agg([sum])
thriller = nf1.groupby(['Year']).Thriller.agg([sum])
comedy = nf1.groupby(['Year']).Comedy.agg([sum])
romance = nf1.groupby(['Year']).Romance.agg([sum])
adventure = nf1.groupby(['Year']).Adventure.agg([sum])
crime = nf1.groupby(['Year']).Crime.agg([sum])
```

```python
# Change the name of the column so it would be more easy to read it 
action = action.rename(columns={'sum': 'Action'})
drama = drama.rename(columns={'sum': 'Drama'})
horror = horror.rename(columns={'sum': 'Horror'})
thriller = thriller.rename(columns={'sum': 'Thriller'})
comedy = comedy.rename(columns={'sum': 'Comedy'})
romance = romance.rename(columns={'sum': 'Romance'})
adventure = adventure.rename(columns={'sum': 'Adventure'})
crime = crime.rename(columns={'sum': 'Crime'})
```

```python
# I merge all the dataframes
data = pd.merge(action, drama, left_on='Year', right_on='Year')
data2 = pd.merge(data, horror, left_on='Year', right_on='Year')
data3 = pd.merge(data2, thriller, left_on='Year', right_on='Year')
data4 = pd.merge(data3, comedy, left_on='Year', right_on='Year')
data5 = pd.merge(data4, romance, left_on='Year', right_on='Year')
data6 = pd.merge(data5, adventure, left_on='Year', right_on='Year')
data7 = pd.merge(data6, crime, left_on='Year', right_on='Year')
data7
```

```python
# Using line chart I visualize the data
plt.figure(figsize=(14,6))
plt.title("Popularity of movie genres 2006-2016")
sns.lineplot(data=data7, dashes=False)
```

```python
# I normalize all the rows so i could see the procentage value of how ofen the genre repeated every year 
omg=data7.div(data7.sum(axis=1), axis=0) * 100
omg
```

```python
# Using line chart I visualize the data
plt.figure(figsize=(14,6))
plt.title("Popularity of movie genres 2006-2016")
sns.lineplot(data=omg, dashes=False)
```

```python
# I vizualize the data with heatmap
plt.figure(figsize=(14,7))
plt.title("Popularity of movie genres 2006-2016")
sns.heatmap(data=omg, annot=True)
plt.xlabel("Genre")
```
