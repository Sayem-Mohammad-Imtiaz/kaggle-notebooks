```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
data=pd.read_csv('/kaggle/input/salary/Salary.csv')
data.head()
```

```python
x=np.array(data['YearsExperience'])
x=x.reshape(-1,1)
y=np.array(data['Salary'])
y=y.reshape(-1,1)
data.head()
```

```python
data.isnull().values.any()
```

```python
data.dtypes
```

```python
data.describe()
```

```python
sns.scatterplot(x='YearsExperience',y='Salary',data=data)
```

```python
data_corr=data.corr()
sns.heatmap(data_corr,cmap='Wistia',annot=True)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
lm_reg=LinearRegression()
lm_reg.fit(x_train,y_train)
predict=lm_reg.predict(x_test.reshape(-1,1))
predict
```

```python
theta=[lm_reg.intercept_]+list(lm_reg.coef_)
theta
```

```python
print("sum of square error:",np.mean((lm_reg.predict(x_test.reshape(-1,1))-y_test)**2))
```

```python
print("The value of r_square:" , lm_reg.score(x_test.reshape(-1,1),y_test))
```

```python
plt.scatter(x_test,y_test,c='black')
plt.plot(x_train,lm_reg.predict(x_train))
plt.xlabel("Number of Years")
plt.ylabel("salary")
plt.title("Visualising test set result")
```

```python
sns.regplot(y_test,predict)
plt.title("Actual vs Predicted")
```

```python
y_mat=y_test.reshape(-1,1)
y_mat
```

```python
dframe=pd.DataFrame({'actual':y_mat.flatten(),'Predicted':predict.flatten()})
dframe.head(10)
```

```python
graph=dframe.head(10)
graph.plot(kind='bar')
plt.title('Actual vs Predicted')
plt.ylabel('Salary')
```

```python
##Check for residual normality and mean
f1=plt.figure(figsize=(12,5))
ax=f1.add_subplot(122)
sns.distplot((y_test-predict),ax=ax,color='b')
ax.axvline((y_test-predict).mean(),color='k',linestyle='--')
ax.set_title('Check for Residual normality & mean: \n Residual eror')
```

The residual mean is zero and residual error plot right skewed

```python
# Check for Multicollinearity
#Variance Inflation Factor
R_square_sk=lm_reg.score(x_test,y_test)
VF = 1/(1- R_square_sk)
VF
```

Variance inflation factor value is more than 5, so there is multicollearity.

THE ACCURACY FOR the model came out to be 97% So model looks pretty good.
