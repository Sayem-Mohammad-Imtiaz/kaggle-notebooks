Reading the data file,

```python
sales = pd.read_csv('../input/nutritional-case-multiple-regression/Peptamant P Sales per Country.csv', header=1, sep=';', index_col=False)
#i am going to creae a copy for later use
sales1 = sales.copy()
print(sales.head())
```

We cannot use this data as the decimal numbers have a ',' instead of a '.'  
So, we need to replace te comma with full stop to convert into float.

```python
sales = sales.stack().str.replace(',', '.').unstack()
```

Doing this will turn the values of 'Observtion' and 'Sales per Capita' into null.  
So we need to replace those columns with the original.   

```python
sales.iloc[:, 0] = sales1.iloc[:, 0]
sales.iloc[:, 5] = sales1.iloc[:, 5]
```

Every column name has a line break(\n). We need to remove it for accessing those columns.

```python
sales.columns = sales.columns.str.replace('\n', ' ')
```

Finally, we need to convert numeric data to float.

```python
sales[['Population (M)', 'Peptamant Plus Sales (M�)', 'Sales per Capita (�)', 'Unemployment rate (% of GNP)', ' Healthcare spending (% of GNP)']] = sales[['Population (M)', 'Peptamant Plus Sales (M�)', 'Sales per Capita (�)', 'Unemployment rate (% of GNP)', ' Healthcare spending (% of GNP)']].astype(float)
```
