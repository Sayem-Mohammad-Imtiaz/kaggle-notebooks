```python
import pandas as pd
import seaborn as sns
import tensorflow as tf
```

# **Visualizing frequency of each Feature**

```python
def value_counts(variable):
    x=data[variable]
    xcount=x.value_counts()
    plt.Figure(figsize=(1,1))
    sns.barplot(x=xcount.index,y=xcount.values)
    plt.xlabel(variable)
    plt.ylabel('frequency')
    plt.show()
    
for i in data.columns:
    value_counts(i)
```

# **Creating dummy variable for each category**

```python
Dummies=pd.get_dummies(data[['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
       'stalk-surface-below-ring', 'stalk-color-above-ring',
       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
       'ring-type', 'spore-print-color', 'population', 'habitat']],drop_first=True)
```

```python
Dummies.head()
```

# **Converting target class to numeric**

```python
target=pd.get_dummies(data['class'],drop_first=True)
```

```python
target.head()
```

# **Creating features and target data set**

```python
x=Dummies
```

```python
y=target
```

# **Spliting train and test dataset**

```python
from sklearn.model_selection import train_test_split
```

```python
xtr,xte,ytr,yte=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
xtr.shape
```

# **Scaling the data**

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
xtr=sc.fit_transform(xtr)
xte=sc.fit_transform(xte)
```

# **Initializing model**

```python
from sklearn.ensemble import RandomForestClassifier
rfc=RandomForestClassifier(n_estimators=150,criterion='entropy')
rfc.fit(xtr,ytr)
```

# **Predicting train data**

```python
yptr=rfc.predict(xtr)
```

```python
from sklearn.metrics import classification_report
cr=print(classification_report(ytr,yptr))
```

```python
ypte=rfc.predict(xte)
```

```python
cr=print(classification_report(yte,ypte))
```

# **We've got a 100% accuracy!**

```python
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Dense(units=100,activation='sigmoid',input_shape=(95,)))
model.add(tf.keras.layers.Dense(units=200, activation='sigmoid'))
model.add(tf.keras.layers.Dense(units=300, activation='sigmoid'))
model.add(tf.keras.layers.Dense(units=200, activation='sigmoid'))
model.add(tf.keras.layers.Dense(units=1, activation='linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer='Adam',loss='mean_squared_error')
```

```python
hist=model.fit(xtr,ytr,epochs=100,batch_size= 50, validation_split=0.23)
```

```python
hist.history.keys()
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
```
