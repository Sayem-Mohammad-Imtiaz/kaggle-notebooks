```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../input/usa-housing/USA_Housing.csv")
```

```python
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(df.corr(),annot=True,cmap=cmap)
```

```python
X=df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
y=df['Price']
```

```python
y.shape
```

```python
X.shape
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python

pd.DataFrame([model.coef_],columns=X.columns).T
```

```python
pred=model.predict(x_test)
```

```python
pred.shape
```

```python
sns.scatterplot(x=pred,y=y_test)
```

```python
from sklearn import metrics
```

```python
print(metrics.mean_squared_error(pred,y_test))
print(metrics.mean_absolute_error(pred,y_test))
print(np.sqrt(metrics.mean_squared_error(pred,y_test)))
```
