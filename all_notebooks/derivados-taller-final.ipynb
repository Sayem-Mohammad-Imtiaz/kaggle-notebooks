```python
import seaborn as sns
data = pd.read_csv("../input/london-bike-sharing-dataset/london_merged.csv")
data.head()
```

En el siguiente cluster de graficas podemos analisar y sacar varias conclusiones:
*  t1 y t2 tieen comportamientos muy parecidos, esto es explicado por quw uno es la tempreatura y el otro la temperatura percibida.
*  De igual manera podemos ver que las variables "is_weekend" e "is_holidays" se comportan de forma similar, y en algunos casos carece de sentido
*  La variable cnt se relaciona con t1, t2, hum y wind_speed SE DISTRIBUYE DE FORMA SIMILAR, pues, tiene una distribucion normal. Esto quiere decir, que a poca o mucha temperatura, real o percivida, las personas alquilan menos bicicletas. Ademas, entre mas humedad, mas personas alquilan bicicletas.


```python
import matplotlib.pyplot as plt
data_sample = data.sample(1000)

p = sns.PairGrid(data=data_sample, vars=['t1', 't2', 'hum', 'wind_speed', 'weather_code', 'is_holiday', 'is_weekend','season', 'cnt'])
p.map_diag(plt.hist)
p.map_offdiag(plt.scatter)
```

```python
#La distribucion de la temperatura, nos da luces sobre como la temperatura de 10 a 20 grados, existe la mayor cantidad de alquileres, cuando la temperatura desciende mucho, de igual manera los alquileres. Por otro lado, si la temperatura aumenta demasiado, igualmente el alquiler cae.
```

* Las siguientes graficas nos muestran correlacion
1) Velocidad del viento y numero de alquileres: se puede ver que entre mucho mas viento, los alquileres disminuyen
2) Humedad y numero de alquileres: tiene una distribución inversa al literal anterior, entre menos humedad, menos alquileres.


```python

plt.figure(figsize=[20,10])
plt.title('Correlación entre velocidad del viento y numero de alquileres')
sns.scatterplot(data['wind_speed'],data['cnt'])
plt.figure(figsize=[20,10])
plt.title('Correlación entre humedad y numeros de alquileres')
sns.scatterplot(data['hum'],data['cnt'] )
```

Para finalizar, observamos la distribucion de dias festivos y fines de semana y se puede ver que las personas que mas usan las bicicletas alquilasdas lo hacen en dias entre semana que no son festivos.
