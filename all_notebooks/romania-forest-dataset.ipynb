```python
forest = pd.read_csv("/kaggle/input/forest-surfaces-evolution-in-romania-19902019/forest_surfaces_romania_en_1990_2019.csv")
```

```python
forest
```

```python
unique_regions = pd.unique(forest["Region"])
unique_regions
```

```python
region_df = forest.groupby(["Region"]).sum()
region_df.drop("Year", inplace=True, axis=1)
region_df.head()
```

```python
sorted_regions_df = region_df.sort_values("Value", ascending=False)
top_10_regions_df = sorted_regions_df.iloc[1:11]

top_10_regions_df.plot(kind="pie", subplots=True, figsize=(10,10), autopct="%.2f", legend=False)
```

```python
regions_values = []

for region in unique_regions:
    region_df = forest.loc[forest['Region'] == region]
    grouped_df = region_df.groupby(['Year'])[['Value']].mean()
    
    for year in list(grouped_df.index):
        year_value = grouped_df['Value'][year]
        regions_values.append([region, year, year_value])
```

```python
regions_values_df = pd.DataFrame(regions_values, columns=['Region', 'Year', 'Value'])
```

```python
sorted_regions = regions_values_df.loc[regions_values_df['Year'] == 2019].sort_values('Value', ascending=False)
sorted_regions.head()
```

```python
def get_slice(start, end):
    return sorted_regions.iloc[start:end+1]['Region'].to_list()
```

```python
def get_slice_arr(slice_df):
    largest_arr = []

    for largest in slice_df:
        largest_df = regions_values_df.loc[regions_values_df['Region'] == largest]

        for year in largest_df['Year']:
            value = largest_df.loc[largest_df['Year'] == year]['Value'].values[0]
            largest_arr.append([largest, year, value])
    
    return largest_arr
```

```python
def get_df_from_slice(arr):
    df = pd.DataFrame(arr, columns=['Region', 'Year', 'Value'])
    df = df.pivot(index='Year', columns='Region', values='Value')
    return df
```

```python
forest_slice = get_slice(15, 24)
forest_slice_arr = get_slice_arr(forest_slice)

df = get_df_from_slice(forest_slice_arr)
df.plot(ylabel="Hectares in thousands", figsize=(10,10))
```
