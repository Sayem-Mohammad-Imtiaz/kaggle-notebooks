```python
df=pd.read_csv('../input/studentsperformance/StudentsPerformance.csv')
```

# To find out total percentage of Marks obtained by Students

```python
df['total score']=df['math score']+df['reading score']+df['writing score']
df['percentage']=round((df['total score']/300)*100,2)
df
```

# Total Number of Female vs Male

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.pie(x=[len(df[df['gender']=='male'])/len(df)*100,len(df[df['gender']=='female'])/len(df)*100],labels=['Male','Female'],autopct='%.2f%%',startangle=90)
```

# We can see that female parent's education level is comparetively better as compared to males except in high shool

```python
plt.figure(figsize=(12,5))
sns.countplot(x=df['parental level of education'],hue=df['gender'])
plt.xticks(rotation=45)
```

# Below figure depicts that students who completed TEST COURSE considerable performed well (Observe the mean in both cases).

```python
import matplotlib.pyplot as plt
import seaborn as sns
data=df[df['test preparation course']=='completed'].sort_values(by='percentage',ascending=False)
data1=df[df['test preparation course']=='none'].sort_values(by='percentage',ascending=False)
data_mean=df[df['test preparation course']=='completed']
data_mean['percentage'].mean()
data1_mean=df[df['test preparation course']=='none']
data1_mean['percentage'].mean()
fig,ax=plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
ax[0].set_title('Complete Test Course')
ax[1].set_title('Non-Completion of Test Course')
sns.histplot(ax=ax[0],x=data['percentage'],bins=10,data=df,kde=True)
sns.histplot(ax=ax[1],x=data1['percentage'],bins=10,data=df,kde=True)
ax[0].axvline(data_mean['percentage'].mean(), color='r', linestyle='--', label="Mean")
ax[1].axvline(data1_mean['percentage'].mean(), color='r', linestyle='--', label="Mean")
```

# Parental Education is higher in "SOME COLLEGES" category as compared to other categories.

```python
sns.countplot(data=df,x='parental level of education',order=df['parental level of education'].value_counts(ascending=True).index,color='Yellow')
plt.xticks(rotation=45)
```

# Females performed better than Males (although the number of female students exceed male students by 3.6%)

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df,x='gender',y='percentage')
```
