```python
medianHouseholdIncome2015 = pd.read_csv('../input/MedianHouseholdIncome2015.csv', encoding="windows-1252")
percentagePeopleBelowPovertyLevel = pd.read_csv('../input/PercentagePeopleBelowPovertyLevel.csv', encoding="windows-1252")
percentOver25CompletedHighSchool = pd.read_csv('../input/PercentOver25CompletedHighSchool.csv', encoding="windows-1252")
shareRaceByCity = pd.read_csv('../input/ShareRaceByCity.csv', encoding="windows-1252")
policeKillingsUS = pd.read_csv('../input/PoliceKillingsUS.csv', encoding="windows-1252")
```

```python
percentagePeopleBelowPovertyLevel.poverty_rate.replace("-",0.0,inplace=True)
percentagePeopleBelowPovertyLevel.poverty_rate = percentagePeopleBelowPovertyLevel.poverty_rate.astype(float)
area_list = list(percentagePeopleBelowPovertyLevel["Geographic Area"].unique())
area_proverty_ratio = []
for i in area_list:
    x = percentagePeopleBelowPovertyLevel[percentagePeopleBelowPovertyLevel["Geographic Area"] == i]
    area_proverty_rate = sum(x.poverty_rate)/len(x)
    area_proverty_ratio.append(area_proverty_rate)
data = pd.DataFrame({"area_list":area_list,"area_provert_ratio":area_proverty_ratio})
new_index = data.area_provert_ratio.sort_values(ascending=False).index.values
sorted_data = data.reindex(new_index)
plt.figure(figsize=(15,10))
sns.barplot(x=sorted_data["area_list"],y=sorted_data["area_provert_ratio"])
plt.xticks(rotation = 30)
plt.xlabel("area list")
plt.ylabel("area provert ratio")
plt.title("Poverty Rate of Each State")
plt.show()
```

```python
#Bar Plot
percentOver25CompletedHighSchool.percent_completed_hs.replace("-",0.0,inplace=True)
percentOver25CompletedHighSchool.percent_completed_hs = percentOver25CompletedHighSchool.percent_completed_hs.astype(float)
area_liste = percentOver25CompletedHighSchool["Geographic Area"].unique()
area_rate_liste = []
for i in area_liste:
    x = percentOver25CompletedHighSchool[percentOver25CompletedHighSchool["Geographic Area"] == i]
    area_ratio = sum(x.percent_completed_hs)/len(x)
    area_rate_liste.append(area_ratio)
df = pd.DataFrame({"area_liste":area_liste,"area_rate_gradution":area_rate_liste})
new_index = df.area_rate_gradution.sort_values(ascending=False).index.values
df = df.reindex(new_index)

plt.figure(figsize=(17,11))
sns.barplot(x=df.area_liste,y=df.area_rate_gradution)
plt.xticks(rotation=30)
plt.title("High School Gradution Rate")
plt.xlabel("Area")
plt.ylabel("Area Rate Gradution")
plt.show()
```

```python
# Bar Plot
shareRaceByCity["share_white"].replace("-",0.0,inplace=True)
shareRaceByCity["share_white"].replace("(X)",0.0,inplace=True)
shareRaceByCity["share_black"].replace("-",0.0,inplace=True)
shareRaceByCity["share_black"].replace("(X)",0.0,inplace=True)
shareRaceByCity["share_native_american"].replace("-",0.0,inplace=True)
shareRaceByCity["share_native_american"].replace("(X)",0.0,inplace=True)
shareRaceByCity["share_asian"].replace("-",0.0,inplace=True)
shareRaceByCity["share_asian"].replace("(X)",0.0,inplace=True)
shareRaceByCity["share_hispanic"].replace("-",0.0,inplace=True)
shareRaceByCity["share_hispanic"].replace("(X)",0.0,inplace=True)
shareRaceByCity.loc[:,["share_white","share_black","share_native_american","share_asian","share_hispanic"]] =shareRaceByCity.loc[:,["share_white","share_black","share_native_american","share_asian","share_hispanic"]].astype(float)
area_listed = shareRaceByCity["Geographic area"].unique()
white = []
black = []
asian = []
american = []
hispanic = []
for i in area_listed:
    x = shareRaceByCity[shareRaceByCity["Geographic area"] == i]
    white.append(sum(x.share_white)/len(x))
    black.append(sum(x.share_black)/len(x))
    american.append(sum(x.share_native_american)/len(x))
    asian.append(sum(x.share_asian)/len(x))
    hispanic.append(sum(x.share_hispanic)/len(x))

f,ax = plt.subplots(figsize=(9,9))
sns.barplot(x=white,y=area_listed, color="g",alpha=0.7)
sns.barplot(x=black,y=area_listed, color="r",alpha=0.7)
sns.barplot(x=american,y=area_listed, color="b",alpha=0.7)
sns.barplot(x=asian,y=area_listed, color="cyan",alpha=0.7)
sns.barplot(x=hispanic,y=area_listed, color="brown",alpha=0.7)
ax.legend(loc="upper right", frameon=True)
plt.show()
```

```python
# Point Plot
sorted_data["area_provert_ratio"] = sorted_data.area_provert_ratio/max(sorted_data.area_provert_ratio)
df["area_rate_gradution"] = df.area_rate_gradution/max(df.area_rate_gradution)
df_concat = pd.concat([sorted_data,df.area_rate_gradution],axis=1)
df_concat.sort_values("area_provert_ratio",inplace=(True))

plt.figure(figsize=(23,9))
sns.pointplot(data=df_concat, x="area_list", y="area_provert_ratio", color="g")
sns.pointplot(data=df_concat, x="area_list", y="area_rate_gradution", color="r")
plt.xlabel("Status", color="b", fontsize=15)
plt.ylabel("Values", color="b", fontsize=15)
plt.text(40,0.55, "Proverty Rate", color="g",fontsize=15)
plt.text(40,0.60, "High School Graduate Rate", color="r",fontsize=15)
plt.title("High School Graduate Rate vs Proverty Rate", fontsize= 27)
plt.grid()
plt.show()
```

```python
# Joint Plot
sns.jointplot("area_provert_ratio","area_rate_gradution", data=df_concat, size=7, kind="kde")
plt.show()
```
