# The expected minimal allocation on key priorities

![prioritization](https://ec.europa.eu/info/sites/default/files/2021feb_rrf_green_digital_wo_hashtags_1.jpg)

**NOTE:** as the amounts could evolve, added a section as a point of reference, so that any variation can be applied directly here

```python
reference_grants = 312.5
reference_loans = 360
reference_total = reference_grants + reference_loans
reference_total
```

```python
endorsed_grants = round(data['ap_grant'].sum(),2) 
endorsed_grants
```

```python
endorsed_loans = round(data['ap_loan'].sum(),2) 
endorsed_loans
```

```python
percentage_endorsed_grants = round(endorsed_grants / reference_grants,2) * 100
percentage_endorsed_grants
```

```python
percentage_endorsed_loans = round(endorsed_loans / reference_loans,2) * 100
percentage_endorsed_loans
```

# The minimal distribution on key priorities expected

![distribution](https://ec.europa.eu/info/sites/default/files/2021feb_rrf_green_digital_wo_hashtags_1.jpg)

```python
average_climate = data['ap_climatepcnt'].sum() / data['ap_climatepcnt'].count()
print ("average allocation to climate: ", average_climate)
```

```python
average_digitalisation = data['ap_digitalisationpcnt'].sum() / data['ap_digitalisationpcnt'].count()
print ("average allocation to digital transformation: ", average_digitalisation)
```
