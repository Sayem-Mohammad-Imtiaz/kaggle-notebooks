```python
import numpy as np
import pandas as pd 

%matplotlib inline
pd.set_option('display.max_rows', 20)
```

Et uurida edasisi andmeid, impordime datasetist ühe csv faili, mis sisaldab infot IMDB-s leiduva kohta.
Anname sellele muutujale nimeks sisend.

```python
sisend = pd.read_csv("../input/IMDB-Movie-Data.csv")
```

Siit leiame, et enim filme IMDB keskkonda sisestatult on loomulikult viimastest aastatest. Käesoleva andmestiku alusel on enim filme linastunud 2016. aastal (297 filmi) ning järjest kahanevalt eelmistel aastatel on linastunud 2015 - 127 filmi, 2014 - 98 filmi, 2013 - 91 filmi jne.

```python
filmid_aastas = sisend["Year"].value_counts();
sisend.Year.plot.hist(bins=len(filmid_aastas), grid=True, rwidth=0.95);
```

Suuremas perspektiivis on iga linastuva filmi idee meelitada ühiskonda kritiseeriva või puhtalt kommertsliku filmiga inimesi kinno, et tulu teenida. Suurfilmide produktsiooni olulisimaks turundusvahendiks on loomulikult inimesed, kes filme vaatamas käivad. Kinolinale hea filmi tuues on küllaltki suur võimalus, et isegi esmapilgul tundmatut filmi võivad inimesed oma sõpruskonnas reklaamida ning seeläbi sellele reklaami teha.

Käesolevaga uurimegi käesoleva andmestiku põhjal, kas ning millises sõltuvuses on filmide tulud võrreldes IMDB-s antud hindega. 

Korrelatsioonigraafikult näeme esiteks, et filmidele antud hinded jäävad üldiselt vahemikku 5 punkti kuni 8 punkti. Samuti on näha tihendamat horisontaalset lineaarset puntkide kogu graafiku alumises osas, mis näitab, et ka kõrgema hindega filmid (8 punkti) võivad teenida samapalju tulu (võrreldes filmidega, mis on saanud vaid 5-6 punkti). Siit paistab ka välja, et suurema osa filmide tulu jääb alla 200 miljoni piiri. Samas on näha kerget diagonaalsuunalist trendi, kus rohkem teenivad filmid on saanud ka IMDB-s parema hinde.

```python
#filmid_2016 = sisend[sisend["Year"]==2016]
filmid_aastas = sisend;
filmid_aastas.plot.scatter("Rating", "Revenue (Millions)", alpha=0.4);
```

Lõpetuseks on põnev teada nende õnnestunud kinofilmide režissööre, kes on olnud produktsioonide tõelised hinged, kelleta poleks filmid kaugeltki sellised tulnud. Viimases osas oleme järjestanud TOP100 režissööri nende filmide keskmise tulu järgi.

```python
pd.set_option('display.max_rows', 100)
režissööride_tulu = sisend.groupby("Director")["Revenue (Millions)"].mean()
režissööride_tulu = režissööride_tulu.sort_values(ascending=False).head(100)
režissööride_tulu
```
