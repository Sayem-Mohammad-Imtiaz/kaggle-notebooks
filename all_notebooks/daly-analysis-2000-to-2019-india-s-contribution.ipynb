# First, I looked into the outline by chart and mapping.

```python
df.groupby('YEAR')['DEATHS'].sum().plot.bar()
plt.title('DEATHS')
```

```python
df.groupby('YEAR')['DALY'].sum().plot.bar()
plt.title('DALY')
```

```python
sns.heatmap(df.corr(),annot=True)
```

```python
pd.DataFrame(df.groupby(['YEAR','COUNTRY_CODE'])['DEATHS'].sum())
```

```python
df_00_19=df[df['YEAR'].isin([2000,2019])]
df1_d=pd.DataFrame(df.groupby(['YEAR','COUNTRY_CODE'])['DEATHS'].sum())
df1_d=df1_d.reset_index()
df1_da=pd.DataFrame(df.groupby(['YEAR','COUNTRY_CODE'])['DALY'].sum())
df1_da=df1_da.reset_index()
```

```python
import matplotlib.pyplot as plt
import plotly_express as px
fig = px.choropleth(df1_d,locations='COUNTRY_CODE',color='DEATHS',scope='world',color_continuous_scale=px.colors.sequential.GnBu,
                    animation_frame='YEAR',range_color=(50000,10000000),title='Top 25 Economics Countries Death Numbers',height=800
    )
fig.show()
```

```python
fig = px.choropleth(df1_da,locations='COUNTRY_CODE',color='DALY',scope='world',color_continuous_scale=px.colors.sequential.GnBu,
                    animation_frame='YEAR',range_color=(150000,5000000),title='Top 25 Economics Countries DALY Numbers',height=800
    )
fig.show()
```

```python
plt.figure(figsize=(10,20))
df_00_19.groupby(['COUNTRY','YEAR'])['DALY'].sum().plot.barh(color=['red',"blue"])
```

# India has lots of DALYS in 25 countries. On the other hand, the deceasing of it between 2000 and 2019 is also very huge.

```python
df_00_19.groupby(['GHE_CAUSE_TYPE','YEAR'])['DALY'].sum().plot.barh(color=['red',"blue"])
```

# Comparing between 2000 and 2019, Communicable, maternal, perinatal and nutritional conditions has been improved. On the other hand, Noncommunicable diseases has been increasing.

# And I can find the causes of increasing and decreasing in below chart.

```python
plt.figure(figsize=(10,20))
df_00_19.groupby(['GHE_CAUSE_TITLE','YEAR'])['DALY'].sum().head(60).plot.barh(color=['red',"blue"])
```

```python
df_00_19.groupby(['SEX_CODE','YEAR'])['DALY'].sum().plot.barh(color=['red',"blue"])
```

# Both male and female, DALY has been decreasing between 2000 and 2019.

```python
plt.figure(figsize=(10,20))
df_00_19.groupby(['AGEGROUP_CODE','YEAR'])['DALY'].sum().plot.barh(color=['red',"blue"])
```

# Especially years 0 to 4, DALY has been improved very much.

# As a next step, I focused on India, where has lots of DALYS in 25 countries, on the other hand, the deceasing of it between 2000 and 2019 is also huge.

```python
df_ind=df[df['COUNTRY']=='India']
```

```python
df_ind.groupby('YEAR')['DALY'].sum().plot.barh()
```

# In India, DALY has been decreasing since 2000.

```python
plt.figure(figsize=(10,20))
df_ind.groupby(['AGEGROUP_CODE','YEAR'])['DALY'].sum().plot.barh(color=['red','blue','yellow','pink','green'])
```

# In India, years 0 to 4 DALY has also been improved very much. It has cotributed to the improvement of global DALY improvement.

```python
plt.figure(figsize=(10,20))
df_ind.groupby(['AGEGROUP_CODE','YEAR'])['DEATHS_100K'].sum().plot.barh(color=['red','blue','yellow','pink','green'])
```

# We can also find the improvement of DEATHS_100k of year 0-1 in India.
