```python
from sklearn.preprocessing import normalize
from sklearn.cluster import KMeans
```

# Clustering en 2 clusters sur toutes les chansons

```python
df=pd.read_csv('../input/spotify-dataset-19212020-160k-tracks/data.csv')
df.head()
```

On se débarasse des caractéristiques qui ne sont pas numériques mais aussi de celle qui ne sont pas pertinentes comme l'année, la durée, id etc...

```python
df1=df.drop(['artists','id','name','year','release_date','duration_ms'],axis=1)
```

```python
tab1 = normalize(df1, axis=0, norm='max')
```

```python
kmeans1 = KMeans(
    init="random",
    n_clusters=2,
    n_init=10,
    max_iter=300,
    random_state=42
)
```

```python
kmeans1.fit(tab1)
kmeans1.inertia_
```

## Clustering en 6 clusters sur une partie des chansons

J'importe les mêmes données que pour le clustering hiérarchique

```python
dfpetit=pd.read_csv('../input/dfpetit2/dfpetit2.csv')
```

```python
dfpetit
```

```python
dfpetit2=dfpetit.drop(['artists','id','name','year','release_date','duration_ms','Unnamed: 0'],axis=1)
```

```python
dfpetit2
```

```python
tab2 =normalize(dfpetit2, axis=0, norm='max')
tab2
```

```python
kmeans2 = KMeans(
    init="random",
    n_clusters=6,
    n_init=10,
    max_iter=300,
    random_state=42
)
```

```python
kmeans2.fit(tab2)
kmeans2.inertia_
```

```python
liste_labels=list(kmeans2.labels_)
```

```python
dfpetit['label']=liste_labels
```
