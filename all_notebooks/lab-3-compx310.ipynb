```python
dataframe = pd.read_csv('/kaggle/input/wisconsin-breast-cancer-cytology-features/wisconsin_breast_cancer.csv')
dataframe=dataframe.fillna(0)
dataframe.info()
```

```python
X= dataframe.iloc[:,1:-1]
X
Y= dataframe.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y, test_size =0.2, random_state=1506360)
```

```python
from sklearn.linear_model import SGDClassifier

sgd_clf =SGDClassifier(random_state=100)
sgd_clf.fit(X_train, Y_train)
```

```python
def nonempty_subsets(lst):
     n = len(lst)
     for i in range(1, 1<<n):
             yield [lst[j] for j in range(n) if ((1 << j) & i) > 0]
```

```python
from sklearn.model_selection import cross_val_score
cols = X.columns.tolist()
subset_name = []
subset_cv = []

for subset in nonempty_subsets(cols):
    CVS = np.mean(cross_val_score(sgd_clf, X_train[subset], Y_train, cv=10, scoring = "accuracy"))
    subset_name.append(subset)
    subset_cv.append(CVS)
   
 
index = subset_name[subset_cv.index(np.max(subset_cv))]  
max_CVS = np.max(subset_cv)
print(max_CVS, index)
```

```python
from sklearn.metrics import accuracy_score
accuracy_num= []
for subset in nonempty_subsets(cols):
    x_sub_train = X_train[subset]
    x_sub_test =X_test[subset]
    sgd_clf.fit(x_sub_train,Y_train)
    y_pred = sgd_clf.predict(x_sub_test)
    accuracy_score(Y_test, y_pred)
    accuracy_num.append(accuracy_score)
    
index = accuracy_num[subset_cv.index(np.max(subset_cv))]
max_CVS = np.max(subset_cv)
print(index)
```
