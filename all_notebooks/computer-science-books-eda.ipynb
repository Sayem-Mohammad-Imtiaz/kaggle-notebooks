```python
data=pd.read_csv("/kaggle/input/top-270-rated-computer-science-programing-books/prog_book.csv")
data.head(5)
```

```python
print(data.info())
```

```python
(data.describe(include='all'))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,10))
plot = sns.scatterplot(x="Reviews",y="Rating",data=data)
plot.set_xticklabels(plot.get_xticklabels(),rotation=90)

plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
wocl=WordCloud(stopwords=STOPWORDS).generate(" ".join(data["Book_title"].tolist()))
plt.figure(figsize=(25,10))
plt.imshow(wocl)
```

```python
sns.pairplot(data)
```

```python
plt.figure(figsize=(19,15))
sns.distplot(data['Rating'],color='red')
plt.title('Rating Distribution')
```

```python
plt.figure(figsize=(19,15))
sns.distplot(data['Price'],color='red')
plt.title('Price Distribution')
```

```python
plt.figure(figsize=(19,15))
sns.boxplot(data=data,y='Price',x='Type')
plt.title('Distribution of Prices by Different Book Types')
```

```python
data['NumberWords'] = data.Book_title.apply(lambda x: len(x.split(' ')))
```

```python
plt.figure(figsize=(19,15))
sns.boxplot(data=data,y='NumberWords',x='Type')
plt.title('Distribution of Words Conut In Book Title And Different Book Types')
```

```python
plt.figure(figsize=(19,15))
sns.jointplot(data=data,y='NumberWords',x='Price',height=10)
plt.title('Distribution of Words Conut In Book Title And Price')
```
