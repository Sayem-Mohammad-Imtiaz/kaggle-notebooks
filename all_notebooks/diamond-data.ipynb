```python
# Problem statement
'''Analyze diamonds by their cut, color, clarity, price, and other attributes'''
```

```python
# importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# improting machine learning model
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
# Reading the data
df = pd.read_csv('../input/diamonds/diamonds.csv')
df.head()
```

```python
# Getting the info of columns
df.info()
```

```python
# checking for null values
df.isna().sum()
```

```python
# dropping the first column as it is not required
df.drop(df.columns[0],axis = 1,inplace = True)
```

```python
# setting target variable(y) and input variable(X)
y = df['price']
X = df.drop('price',axis = 1)
```

```python
# checking number of unique items
print(f"cuts: {len(df['cut'].unique())}")
print(f"colors: {len(df['color'].unique())}")
print(f"clarities: {len(df['clarity'].unique())}")
```

```python
# cooncoding each catogorical value to numeric
encoder  = LabelEncoder()

X['cut'] = encoder.fit_transform(X['cut'])
cut_map = {index: label for index, label in enumerate(encoder.classes_)}

X['color'] = encoder.fit_transform(X['color'])
color_map = {index: label for index, label in enumerate(encoder.classes_)}

X['clarity'] = encoder.fit_transform(X['clarity'])
clarity_map = {index: label for index, label in enumerate(encoder.classes_)}

```

```python
print(cut_map)
print(color_map)
print(clarity_map)
```

```python
# modified data 
X
```

```python
# normalisation
scaler = MinMaxScaler()

X = scaler.fit_transform(X)
pd.DataFrame(X)
```

```python
# training model
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.8)
```

```python
# model 1 
lr = LinearRegression()
#model 2
las_lr = Lasso(alpha = 0.1)
# model 3 
rid_lr = Ridge(alpha = 0.1)

lr.fit(X_train,y_train)
las_lr.fit(X_train,y_train)
rid_lr.fit(X_train,y_train)
```

```python
# print scores of all three models
print(f" without regularization: {lr.score(X_test,y_test)}")
print(f" Lasso regularization: {las_lr.score(X_test,y_test)}")
print(f" Ridge regularization: {rid_lr.score(X_test,y_test)}")
```
