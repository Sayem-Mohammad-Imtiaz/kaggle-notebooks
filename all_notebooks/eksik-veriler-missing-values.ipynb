# Eksik Veriler



```python

import numpy as np 

import pandas as pd

```

```python
data=pd.read_csv('../input/videogamesales/vgsales.csv')
```

Veriler hakkında bilgi edinmek için info() fonksiyonunu kullandık.Burada Year ve Publisher sütununlarında  eksik veriler olduğunu görüyoruz.

```python
data.info()
```

Tüm sütunlardaki eksik veri sayısını bulmak için pandas kütüphanesinde bulunan isnull() fonksiyonunu kullanıyoruz.


```python
missing_values_count = data.isnull().sum() 

missing_values_count
```

Toplam ne kadar eksik veri olduğunu buluyoruz.

```python
total_missing = missing_values_count.sum()

total_missing
```

 Eksik veri bulunan sütunu silmek için pandas kütüphanesine ait olan dropna() fonksiyonunu kullanıyoruz.

```python
new_data=data.dropna(axis=1)

new_data
```

 Burada eksik veri olan satırları sildik.

```python
new_data=data.dropna()

new_data
```

Bir diğer yöntem ise pandas kütüphanesinde bulunan fillna() fonksiyonu ile eksik verileri istediğimiz değer ile doldurabiliriz.

```python
new_data=data.fillna(0) 

new_data
```

```python
new_data.info()
```

Burada örnek olması bakımından her sütunun ortalama değerini hesaplayıp eksik olan veriler yerine fillna() fonksiyonu ile doldurduk.


```python
mean=data.mean()  
new_data=data.fillna(mean) 
new_data
```

Bir başka yöntem ise replace() fonksiyonu burada ise yukarıda olduğu gibi ortalama değeri hesaplayıp eksik veriler ile yer değiştirme işlemi yapılıyor.

```python
mean=data.mean() 
new_data=data.replace(np.NaN,mean)
new_data
```

Ve son olarak Imputation işlemi yapıyoruz. Sklearn kütüphanesini kullanacağız.

```python

from sklearn.impute import SimpleImputer

imputer=SimpleImputer(missing_values=np.nan, strategy='mean') # ortalama hesaplamak içim

new_data=data.iloc[:,3:4].values # Year sütununu seçiyoruz, diğer sütunların bazıları numerik değerler olmadığı için ortalama hesaplanırken hata alabilriz.


imputer = imputer.fit_transform(new_data)


imputers = pd.DataFrame(data=imputer)

```

```python
missing_values_count = imputers.isnull().sum() 

missing_values_count
```
