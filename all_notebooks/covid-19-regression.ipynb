#confirmed and deaths oranlarını linear regression ile inceleyelim

```python
#genel bir şeklini inceleyelim.
plt.scatter(data.Deaths,data.Confirmed)
plt.xlabel("Confirmed") # Tanı konulan kişi sayısı
plt.ylabel("Deaths") # Ölen  Kişi sayısı
plt.title("ORAN")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
x=data.Confirmed.values.reshape(-1,1)
y=data.Deaths.values.reshape(-1,1)

#fit etme 
lr.fit(x,y)
```

```python
#kesişim
b0=lr.intercept_
print(b0)
```

```python
#egim
b1=lr.coef_
print(b1)
```

```python

y_head=lr.predict(x)
plt.plot(x,y_head,color="red")

plt.show()
```
