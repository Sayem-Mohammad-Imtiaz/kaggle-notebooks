# Part A

```python
df = pd.read_csv("/kaggle/input/house-prices-advanced-regression-techniques/train.csv")
df_t = pd.read_csv("/kaggle/input/house-prices-advanced-regression-techniques/test.csv")
```

```python
## Scikit-Learn library
from sklearn.linear_model import LinearRegression

model = LinearRegression()
## X and y
feature = ['GrLivArea']
target = ['SalePrice']


X_train = df[:1200][feature]
y_train = df[:1200][target]
```

## 1. With the data points, fit a linear model.

```python
model.fit(X_train, y_train)
```

## 2. Draw a scatterplot with the linear model as a line

```python
sns.regplot(x=df['GrLivArea'], y=df['SalePrice']);
```

## 3. Analyze the linear model fitted and examine whether predictor variables seem to have a significant influence on the outcome (No wrong answer here)

```python
x = df['GrLivArea']
y = df['SalePrice']

predict = df['SalePrice'].mean()
errors = predict - df['SalePrice']
mean_absolute_error = errors.abs().mean()

sns.lineplot(x=x, y=predict, color='red')
sns.scatterplot(x=x, y=y, color='blue');
```

Based on the above two graphs (one created by linear regression, other used with only mean() value,) we can see that linear regression model has more accurate line.

## 4. Using the model, predict the outcomes for the same data points 

```python
## All test data
X_test = [[x] for x in df[1001:]['GrLivArea']]
y_pred = model.predict(X_test)
```

```python
## train data
plt.scatter(X_train, y_train, color='blue', linewidth=1)

## test data
plt.plot(X_test, y_pred, color='red', linewidth=1);
```

## 5. Compare the outcomes with the actual values of the data set.

```python
sns.regplot(x=df['GrLivArea'], y=df['SalePrice']);
## train data
plt.scatter(X_train, y_train, color='blue', linewidth=1)

## test data
plt.plot(X_test, y_pred, color='red', linewidth=1);
```

Result: slightly different.

## 6. Use your NumPy and Pandas skills to calculate the sum of squares of residuals for your model

```python
output = [[x] for x in df[1001:]['SalePrice']]
prediction = np.round(y_pred,0).astype(int)

residual = output - prediction
RSS = sum(residual **2)[0]
RSS
```

```python
# Load the dataset
table = pd.read_csv('../input/google-play-store-apps/googleplaystore.csv')
table.head()
```

```python
df = table[['Rating','Reviews', 'Size','Installs', 'Price']]
df.Installs = df.Installs.str.replace(r"[a-zA-Z\+\,]",'').replace('', np.nan).fillna("0").astype(float)
df.Size = df.Size.str.replace(r"[a-zA-Z\+\,]",'').str.strip().replace('', np.nan).fillna("0").astype(float)
df.Reviews = df.Reviews.str.replace(r"[a-zA-Z\+\,]",'').str.strip().replace('', np.nan).fillna("0").astype(float)
df.Price = df.Price.str.replace(r"[a-zA-Z\+\,\$]",'').str.strip().replace('', np.nan).fillna("0").astype(float)
df
```

```python
vals, names, xs = [],[],[]
for i, col in enumerate(df.columns):
    vals.append(df[col].values)
    names.append(col)
    xs.append(np.random.normal(i + 1, 0.04, df[col].values.shape[0]))  # adds jitter to the data points - can be adjusted
```

```python
plt.boxplot(vals, labels=names)
palette = ['r', 'g', 'b', 'y']
for x, val, c in zip(xs, vals, palette):
    plt.scatter(x, val, alpha=0.4, color=c)
plt.show()
```

Observation: seems like there are only free app in this data
