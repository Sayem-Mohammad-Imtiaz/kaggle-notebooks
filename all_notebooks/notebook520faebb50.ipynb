```python
X = data.drop(['Date','Catalyst %', 'Type of Catalist', 'Bench life in Minutes'],axis=1)
y= data['Bench life in Minutes']
```

```python
X.head()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=7)
```

```python
print(X_train.shape)
print(X_test.shape)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(X_train,y_train)
pred = LR.predict(X_test)
```

# Print R-Square Value

```python
from sklearn.metrics import r2_score
R2_score= r2_score(y_test,pred)
print(R2_score)
```

# Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 2)
x_train_poly = poly.fit_transform(X_train)
x_test_poly = poly.fit_transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_train_poly,y_train)
```

```python
reg.score(x_train_poly,y_train)
```

```python
reg.score(x_test_poly,y_test)
```

# Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor
dt_reg = DecisionTreeRegressor(max_depth = 2, random_state = 5)
dt_reg.fit(X_train,y_train)
```

```python
dt_reg.score(X_train,y_train)
```

```python
dt_reg.score(X_test,y_test)
```

# Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(n_estimators = 3, random_state = 0)
rf_reg.fit(X_train, y_train)
```

```python
rf_reg.score(X_train,y_train)
```

```python
rf_reg.score(X_test,y_test)
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha = 1)
ridge_reg.fit(x_train_poly, y_train)
```

```python
ridge_reg.score(x_train_poly,y_train)
```

```python
ridge_reg.score(x_test_poly,y_test)
```

# Lasso regression

```python
from sklearn.linear_model import Lasso
la_reg = Lasso()
la_reg.fit(x_train_poly, y_train)
```

```python
la_reg.score(x_train_poly,y_train)
```

```python
la_reg.score(x_test_poly,y_test)
```
