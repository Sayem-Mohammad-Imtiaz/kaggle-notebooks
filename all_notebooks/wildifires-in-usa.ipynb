Removing useless columns for better analysis.

```python
wf.drop(['Unnamed: 0', 'index'],axis=1,inplace=True)
wf.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,6))
sns.lineplot(x='FIRE_YEAR',y='FIRE_SIZE',data=wf)
plt.title("Wildfire Sizes in USA throughout the years",fontweight="bold",fontsize=20)
```

In the graph above it can be seen that since 2005 firesize of wildfires has increased in USA.


Since the title of our dataset mentioned about only the fire records from 2004 to 2015,we will do extreme data analysis i.e  analyse wildfires from year 2004 and wildfires from 2015

```python
#Making 2 different dataframes for 2004 and 2015 wildfire data for better analysis
wf_2004=wf[wf["FIRE_YEAR"]==2004]
wf_2015=wf[wf["FIRE_YEAR"]==2015]
```

```python
#Basic data cleaning by dropping null values
wf_2004.isnull().sum()
wf_2004.dropna()
wf_2015.isnull().sum()
wf_2015.dropna()
```

# Comparing class fire types in 2004 and 2015


```python
plt.figure(figsize=(12,6))
plt.hist([wf_2004["FIRE_SIZE_CLASS"],wf_2015["FIRE_SIZE_CLASS"]],color=["red","green"],label=["Fires in 2004","Fires in 2015"],bins=20)
plt.title("Class types of Fires in 2004 and 2015",fontweight="bold",fontsize=20)
plt.xlabel("Class Type")
plt.ylabel("No of fires")
plt.legend()
```

> What we can notice from this graph is that,although the lower classes of fires have reduced since 2004,riskier classes of fire have slightly increased in USA in 2015

# Comparing Fire size class in each state of USA in 2004

```python
plt.figure(figsize=(20,6))
sns.countplot(x='STATE', hue="FIRE_SIZE_CLASS", data=wf_2004)
plt.xticks(rotation="vertical")
plt.legend(loc="right")
plt.title("Fires in 2004 in various States of USA with Class type",fontsize=20,fontweight="bold")
plt.xlabel("USA states")
plt.ylabel("Fires")
plt.style.use('ggplot')
```

> As class A fires seem the most popular kind we'll look more into that

* Arizona has the most cases of class A fires in 2004

# Comparing Fire size class in each state of USA in 2015

```python
plt.figure(figsize=(20,6))
sns.countplot(x='STATE', hue="FIRE_SIZE_CLASS", data=wf_2015)
plt.xticks(rotation="vertical")
plt.legend(loc="right")
plt.title("Fires in 2015 in various States of USA with Class type",fontsize=20,fontweight="bold")
plt.xlabel("USA states")
plt.ylabel("Fires")

```

* **California had the most Class A fire type in 2015**

# Taking all the wildfires in 2004 into consideration,trying to establish a relationship between the usual number of days taken to extinguish the fire versus the state in which the fire occured

```python
plt.figure(figsize=(20,6))
sns.scatterplot(x="STATE",y="Days_to_extinguish_fire",data=wf_2004)
plt.title("Checking the usual no of days taken to extinguish wildfires in USA in 2004",fontsize=20,fontweight="bold")
```

**It can be observed that Alaska out of a number of wildfires it faced,took the most time to extinguish wildires in 2004 compared to the other states of USA**



# Taking all the wildfires in 2015 into consideration,trying to establish a relationship between the usual number of days taken to extinguish the fire versus the state in which the fire occured

```python
plt.figure(figsize=(20,6))
sns.scatterplot(x="STATE",y="Days_to_extinguish_fire",data=wf_2015)
plt.title("Checking the usual no of days taken to extinguish wildfires in USA in 2015",fontsize=20,fontweight="bold")
```

**Although Alaska marginally bettered its usual fire extinguishing time period since 2004, many other USA states like Washington and California worsened in this particular aspect in 2015**

Is Industrialization to blame??

# Since Arizona was the leading state with most class A type fires in 2004,we compare the time taken to extinguish a fire versus the fire size in acres in AZ in 2004

```python
plt.figure(figsize=(12,6))
wf_2004_az=wf_2004[wf_2004["STATE"]=="AZ"]

plt.scatter(x=wf_2004_az["Days_to_extinguish_fire"],y=wf_2004_az["FIRE_SIZE"])
plt.title("Days to extinguish fire vs Fire size in Arizona in 2004",fontsize=20,fontweight="bold")
plt.xlabel("Days to extinguish fire")
plt.ylabel("Fire Size")
```

**Most fires were extinguished between the 1st and the 25th day in Arizona in 2004 including some of the larger ones in size,some outliers are present though.**

# Since California was the leading state with most class A type fires in 2015,we compare the time taken to extinguish a fire versus the fire size in acres in CA in 2015

```python
plt.figure(figsize=(12,6))
wf_2015_ca=wf_2015[wf_2015["STATE"]=="CA"]

plt.scatter(x=wf_2015_ca["Days_to_extinguish_fire"],y=wf_2015_ca["FIRE_SIZE"],c="green")
plt.title("Days to extinguish fire vs Fire size in California in 2015",fontweight="bold")
plt.xlabel("Days to extinguish fire")
plt.ylabel("Fire Size")
```

**In 2015 class A fires generally took a far greater time to be extinguished with a cluster of them even taking more than 100 days to get extinguished!**

# Mapping down the longitudes and latitudes of Arizona to roughly find the region of wildfires in 2004

```python
plt.figure(figsize=(6,6))
plt.scatter(x="LONGITUDE",y="LATITUDE",c="FIRE_SIZE",data=wf_2004_az,alpha=0.3)
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Heatmap of Arizona wrt. Fire sizes in 2004",fontsize=20,fontweight="bold")
```

**North western and Central Arizona seems to have experienced the worst(biggest) wildfires in 2004**
