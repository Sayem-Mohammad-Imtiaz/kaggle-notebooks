```python
cd = ['SO2', 'O3',
       'Benzene', 'Toluene', 'Xylene']
```

```python
for i in cd:
    alpha = df[['StationId', 'Datetime', i]]

    for j in alpha.loc[:,'StationId'].unique():
        beta = alpha.loc[alpha.loc[:,'StationId']==j]
        beta.pop('StationId')
        beta.columns = ['ds','y']
        beta.dropna(inplace=True)
        try:
            m = Prophet(weekly_seasonality=False, yearly_seasonality=20)
            m.add_country_holidays(country_name='IN')
            m.add_seasonality(name='monthly', period=30.5, fourier_order=5)

            m.fit(beta)

            name = str(i)+"_"+str(j)+".json"

            with open(name, 'w') as fout:
                json.dump(model_to_json(m), fout)
            print(name)
        except:
            print("ERROR : "+i+"____"+j)
```
