```python
import pandas as pd
import numpy as np
import seaborn as sns 
import math
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
df = pd.read_csv("../input/predictingese/amsPrediction - Sheet1.csv")
df.head()
```

```python
df.describe()
```

```python
corr=df.corr()
corr.style.background_gradient(cmap='coolwarm')
```

```python
endog = df['ESE']
exog = sm.add_constant(df[['MSE','Attendance','HRS']])
print(exog)
```

```python
X=exog.to_numpy()
Y= endog.to_numpy()
s1_xt =np.transpose(X)
print(s1_xt)
```

```python
s2_mul1= np.matmul(s1_xt,X)
print(s2_mul1)
```

```python
s3_inv=np.linalg.inv(s2_mul1)
print(s3_inv)
```

```python
s4_mul= np.matmul(s3_inv,s1_xt)
print(s4_mul)
```

```python
s5_res =np.matmul(s4_mul,Y)
print(s5_res)
```

```python
mod = sm.OLS(endog, exog)
results = mod.fit()
print (results.summary())
```
