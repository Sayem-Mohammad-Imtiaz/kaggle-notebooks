# Exploratory Data Analysis

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title('Distribution plot of Year')
sb.distplot(df.Year)
plt.subplot(2,2,2)
plt.title('Histogram of Year')
df['Year'].hist()
plt.subplot(2,2,3)
plt.title('Distribution plot of Life expectancy')
sb.distplot(df['Life expectancy'])
plt.subplot(2,2,4)
plt.title('Histogram of Life expectancy')
df['Year'].hist()
```

```python
sb.heatmap(df.corr())
```

```python
plt.figure(figsize=(12,5))
sb.scatterplot(df.Year,df['Life expectancy'],hue=df['Entity'])
```

```python
countries=df.Entity.unique()
for i in countries:
    data=df[df.Entity==i]
    plt.xlabel('Year')
    plt.ylabel('Life expectancy')
    plt.title(i)
    plt.plot(data.Year,data['Life expectancy'],c='b')
    plt.show()
```
