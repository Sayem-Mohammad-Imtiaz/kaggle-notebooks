```python
df=pd.read_csv("../input/pokemon-challenge/pokemon.csv")
df.info()
df.describe()
df.head(8)
```

```python
#correlation map
#df.corr() gives me features correlation
f,ax=plt.subplots(figsize=(20,10))
sns.heatmap(df.corr(), annot=True, linewidths=.5, fmt= '.2f',ax=ax)# (if annot=True we can see values)(.1f to set decimal part)
plt.show()
```

```python
# Line Plot
df.Speed.plot(kind = 'line', color = 'green',label = 'Speed',linewidth=2,alpha = 0.5,grid = True,linestyle = ':',figsize=(20,10))
df.Defense.plot(kind="line", color = 'red',label = 'Defense',linewidth=2, alpha = 0.5,grid = True,linestyle = '-.',figsize=(20,10))
plt.legend()    
plt.xlabel('Speed')              
plt.ylabel('Defense')
plt.title('Line Plot')           
plt.show()
```

```python
# Scatter Plot 

df.plot(kind='scatter', x='Attack', y='Defense',alpha = 0.5,color = 'purple',linewidth=10,figsize=(16,10))
plt.xlabel('Attack')             
plt.ylabel('Defence')
plt.legend()
plt.title('Attack Defense Scatter Plot')
```

```python
# Histogram
df.Speed.plot(kind = 'hist',bins = 50,color="yellow",figsize = (16,10)) # bins = number of bar
plt.legend()
plt.xlabel("Speed")
plt.ylabel("Frequency")
plt.show()
#plt.clf() clear the plot
```
