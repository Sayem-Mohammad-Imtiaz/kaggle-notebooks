```python
movie_df = pd.read_csv("./movies.csv")
movie_df.head()
```

```python
movie_df['List of Genres'] = movie_df['genre'].apply(lambda x: x.split(','))
movie_df['List of Genres']
```

```python
Genre2Emotion_map = {
    'anger': ['Family', 'Animation', 'Reality-TV'],
    'disgust': ['Musical', 'Fantasy'],
    'fear': ['Sport', 'Documentary', 'Biography', 'News'],
    'joy': ['Action', 'Adult', 'Romance', 'Thriller', 'Western'],
    'sadness': ['Drama', 'History', 'Music', 'Adventure', 'Comedy'],
    'surprise': ['Sci-Fi', 'War', 'Mystery', 'Crime', 'Film-Noir', 'Horror']
}
```

```python
def Genre2Emotion(genre_list):
    emotion_list = []
    for genre in genre_list:
        genre = genre.strip(' ')
        if genre in Genre2Emotion_map['anger']:
            emotion_list.append('anger')
        if genre in Genre2Emotion_map['disgust']:
            emotion_list.append('disgust')
        if genre in Genre2Emotion_map['fear']:
            emotion_list.append('fear')
        if genre in Genre2Emotion_map['joy']:
            emotion_list.append('joy')
        if genre in Genre2Emotion_map['sadness']:
            emotion_list.append('sadness')
        if genre in Genre2Emotion_map['surprise']:
            emotion_list.append('surprise')
            
    return emotion_list
```

```python
movie_df['List of Emotions'] = movie_df['List of Genres'].apply(Genre2Emotion)
movie_df['List of Emotions']
```

```python
movie_df['anger'] = np.zeros((len(movie_df),1))
movie_df['disgust'] = np.zeros((len(movie_df),1))
movie_df['fear'] = np.zeros((len(movie_df),1))
movie_df['joy'] = np.zeros((len(movie_df),1))
movie_df['sadness'] = np.zeros((len(movie_df),1))
movie_df['surprise'] = np.zeros((len(movie_df),1))
```

```python
# def onehotemotions(emotion_list):
#     if 'anger' in emotion_list:
#         df_train['anger'] = 1
```

```python
for i in ['anger', 'disgust', 'fear', 'joy', 'sadness', 'surprise']:
    movie_df[i] = movie_df['List of Emotions'].apply(lambda x: 1 if i in x else 0)
```
