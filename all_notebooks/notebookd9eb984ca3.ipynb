#**Analyzing the data**#

Quick look at the data, to get an overview

**VARIABLE DESCRIPTIONS:** 

 1.  UNID: ISO numeric country code (used by the United Nations) 
 2. WBID: ISO alpha country code (used by the World Bank) 
 3. SES: Socioeconomic status score (percentile) based on GDP per capita and educational attainment (n=174)
 4. country: Short country name 
 5. year: Survey year 
 6. SES: Socioeconomic status score (1-99) for each of 174 countries 
 7. gdppc: GDP per capital Single time-series (imputed) 
 8. yrseduc: Completed years of education in the adult (15+) population 
 9. popshare: Total population shares
