Identify uniqueness in 'Rocket' and see how we can fill NA values

```python
data[" Rocket"].value_counts()

# Since Rocket signifies cost of a Rocket in Millions, we cannot impute any mean/mode values herewith.
# Instead let us fill these values as '0'

# Replace commas from Column values
data[' Rocket'] = data[' Rocket'].str.replace(",", "")

# Convert Rocket column from str to float
data[' Rocket'] = data[' Rocket'].astype(str).astype(float)

# Lastly, Replace NA values as 0
data[' Rocket'].fillna(0, inplace=True)

data.isna().sum()
```

### Identify which company has the highest cost per Rocket 

NASA has the higest cost per Rocket. It starts from 350 Millions and surges above

```python
plt.figure(figsize=(20, 10))
x = sns.barplot(x=data["Company Name"], y=data[" Rocket"])
plt.title('Cost per Rocket w.r.t Companies')
for item in x.get_xticklabels():
    item.set_rotation(90)
```

1) Private companies have incurred more cost when their rockets have failed than State Run Companies

2) The more the State Run Companies spent on their Rockets, that more is their observed Success Rate in launchinig Rockets

```python
plt.figure(figsize=(10, 5))
x = sns.barplot(x=data["Status Mission"], y=data[" Rocket"], hue=data["Private or State Run"])
plt.title('Cost-Success Ratio of Private & State Run Companies')
plt.show()
```

1) Since the 1990s, Private Sector has launcher more rockets than State run companies

Which Company has the most successful Rocket Launches?

```python
successfulLaunches = []
for i in range(0, data[" Rocket"].count()):
    if data["Status Mission"][i] == "Success":
        successfulLaunches.append(data["Company Name"][i])

plt.figure(figsize=(20, 5))
plt.title("Most Successful Rocket Launches")
x = sns.countplot(successfulLaunches)
for item in x.get_xticklabels():
    item.set_rotation(90)
```

- 1971 was the year which has witnessed highest Rocket Launches till date
- Period between 1980 - 2012 can be observed as a lag period

```python
plt.figure(figsize=(20, 5))
plt.title("Rocket Launches per year")
plt.grid()
x = sns.countplot(data["Year"])
for item in x.get_xticklabels():
    item.set_rotation(90)
```
