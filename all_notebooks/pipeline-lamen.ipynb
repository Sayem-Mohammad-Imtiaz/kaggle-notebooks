O miojo como conhecemos, ou Rámen/Lámen para o resto do mundo, é uma comida rápida e barata, mas de qualidade as vezes duvidosa.

A seguir vamos conferir uma ánalise de lámens de  vários países e vamos decidir qual marca compraremos.

```python
#Importando o pandas, matplotlib e a nossa base de dados em csv fornecida pelo próprio Kaggle
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv("../input/lalamen/ramen-ratings.csv")
#Vamos retirar algumas colunas como a top ten, já que utilizaremos nossa própria metodologia para classificação
df.drop(columns=['Top Ten','Style'], inplace=True)
```

Analisando a fonte de dados notei que a coluna 'Stars' está no formato object o que irá atrapalhar nossas buscas por determinadas estrelas e comparações matemáticas entre elas

```python
df[['Stars']].dtypes
```

Agora precisamos converter o campo para float, mas temos uma string unrated no campo, logo, precisamos remove-la para termos apenas números

```python
df['Stars'] = pd.to_numeric(df['Stars'], errors='coerce')
df.dtypes
```

Agora poderemos considerar todos as variedades e marcas com a nota maior que 4 como sendo um bom lámen

```python
good_lamen = df[(df['Stars'] > 4)]
good_lamen
```

Temos 734 lamens considerados bons o suficiente para experimentarmos! <br>
Precisamos ser mais específicos na nossa busca<br> vamos ver as marcas do país mais famoso pelos lámens, o Japão



```python
good_lamen_japan = good_lamen[(good_lamen['Country'] == 'Japan')]
good_lamen_japan
```

159 continua sendo um número muito alto, ouvi dizer que os lámens de kimchi são bem famosos e apimentados

```python
good_lamen = good_lamen_japan[good_lamen_japan['Variety'].str.contains("Kimchi")==True]
good_lamen
```

Parece que já temos uma escolha mais plausível entre dois itens, mas vamos ter uma visão mais ampla do mercado global

```python
#Vamos agrupar por país para sabermos quais países tem mais variedades de marcas
df2 = df.groupby('Country')['Brand'].nunique()
df2.sort_values(ascending=False, inplace=True)
df3= df2.head(10)
df3.plot(kind='bar')
plt.ylabel('Number of brands')
plt.xlabel('Countries')
plt.show()
```

Mas e quanto a qualidade? Vamos ver agora a classificação por estrelas

```python
df2 = df.groupby('Country')['Stars'].mean()
df2.sort_values(ascending=False, inplace=True)
df3= df2.head(10)
df3.plot(kind='bar')
plt.ylabel('mean value of stars')
plt.xlabel('Countries')
plt.show()
```
