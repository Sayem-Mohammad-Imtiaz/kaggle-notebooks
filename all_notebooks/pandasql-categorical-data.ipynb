```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
print("Setup Complete")
import pandasql as psql
```

```python
df = pd.read_csv('../input/laptop-price/laptop_price.csv',encoding='latin-1')
df.head()
```

```python

df["Ram"] = df["Ram"].str.replace('GB', '')
df["Weight"] = df["Weight"].str.replace('kg', '')
df = df.drop(["laptop_ID"],axis=1)
```

```python
df["Ram"] = df["Ram"].astype(int)
df["Weight"] = df["Weight"].astype(float)
```

```python

df['Company'].value_counts()
fig_dims = (20, 6)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(x="Company", data=df, ax=ax)
```

```python
object_values = df.select_dtypes(include='object').columns
object_values
```

```python
# select unique values
df_Product = df['Product'].unique()
df_Company = df['Company'].unique()
df_TypeName = df['TypeName'].unique()
df_OpSys = df['OpSys'].unique()
df_ScreenResolution = df['ScreenResolution'].unique()
df_Cpu = df['Cpu'].unique()
df_Memory = df['Memory'].unique()
df_Gpu = df['Gpu'].unique()


```

```python
# create dataframe for each one
df_Product = pd.DataFrame(df_Product)
df_Company = pd.DataFrame(df_Company) 
df_TypeName = pd.DataFrame(df_TypeName) 
df_OpSys = pd.DataFrame(df_OpSys) 
df_ScreenResolution = pd.DataFrame(df_ScreenResolution) 
df_Cpu = pd.DataFrame(df_Cpu)
df_Memory = pd.DataFrame(df_Memory)
df_Gpu = pd.DataFrame(df_Gpu)

```

```python
#   rename column
df_Product = df_Product.rename(columns={0: "Products"})
df_Company = df_Company.rename(columns={0: "Company"})
df_TypeName = df_TypeName.rename(columns={0: "TypeName"})
df_OpSys = df_OpSys.rename(columns={0: "OpSys"})
df_ScreenResolution = df_ScreenResolution.rename(columns={0: "ScreenResolution"})
df_Cpu = df_Cpu.rename(columns={0: "Cpu"})
df_Memory = df_Memory.rename(columns={0: "Memory"})
df_Gpu = df_Gpu.rename(columns={0: "Gpu"})
```

```python
#  add primary key
df_Product['Product_id'] = df_Product.index
df_Company['Company_id'] = df_Company.index
df_TypeName['TypeName_id'] = df_TypeName.index
df_OpSys['OpSys_id'] = df_OpSys.index
df_ScreenResolution['ScreenResolution_id'] = df_ScreenResolution.index
df_Cpu['Cpu_id'] = df_Cpu.index
df_Memory['Memory_id'] = df_Memory.index
df_Gpu['Gpu_id'] = df_Gpu.index
```

```python

df_final =  psql.sqldf("select * from df a  "
                  "left outer join df_Product b on  a.Product = b.Products "
                  "left outer join df_Company c on  a.Company = c.Company "
                  "left outer join df_TypeName d on  a.TypeName = d.TypeName "
                  "left outer join df_OpSys e on  a.OpSys = e.OpSys "
                  "left outer join df_ScreenResolution f on  a.ScreenResolution = f.ScreenResolution "
                  "left outer join df_Cpu g on  a.Cpu = g.Cpu "
                  "left outer join df_Memory h on  a.Memory = h.Memory "
                  "left outer join df_Gpu i on  a.Gpu = i.Gpu")
```

```python
df_final.head()
```

```python
df_final_object_features = df_final.select_dtypes(include='object').columns
df_final_object_features
```

```python
df_final=df_final.drop(columns=df_final_object_features,axis=1)
```
