# QQ-Test

```python
plt.figure(figsize=(10,5))
stats.probplot(df['IMDB Score'],dist='norm',plot=plt)
plt.show()
```

here it is not normal curve

# Apply BoxCox Transformation

```python
imdb_t,c=stats.boxcox(df['IMDB Score'])
print(c)
```

```python
plt.figure(figsize=(10,5))
stats.probplot(imdb_t,dist='norm',plot=plt)
plt.show()
```

here it is almost normal distribution

```python
sns.displot(df['Runtime'],kde=True)
```

here this graph is left skewd that mean more outlier is on left side

```python
plt.figure(figsize=(20,5))
sns.boxplot(x=df['Runtime'])
plt.xticks(rotation=90)
plt.show()
```

here is outlier detected which is less than 50 or greater than 140

# QQ-Test

```python
stats.probplot(df['Runtime'],dist='norm',plot=plt)
plt.show()
```

here it is not normally distributed

# Apply BoxCox Transformation

```python
x_t,l=stats.boxcox(df['Runtime'])
print(l)
```

```python
stats.probplot(x_t,dist='norm',plot=plt)
plt.show()
```

here it is almost normalized
