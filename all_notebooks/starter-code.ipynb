```python
data = pd.read_csv("/kaggle/input/instela-entries/temiz_girdiler.csv")
```

## Columns
Dataset contains following 3 colummns:
- `entries.entry` : user created, preprocessed clean content free of punctuations and html tags.
- `entries.vote_positive` : positive votes for entry
- `entries.vote_negative` : negative votes for entry

## Example pipeline to predict positive votes for an entry

Given below is an example pipeline using hash vectorizer and linear regression to predict number of positive votes for an entry.

```python
from sklearn import pipeline, feature_extraction, linear_model
```

```python
p = pipeline.Pipeline([
    ('vectorizer', feature_extraction.text.HashingVectorizer()),
    ('regressor', linear_model.SGDRegressor())
])
```

```python
p.fit(data['entries.entry'].iloc[:9000], data['entries.vote_positive'].iloc[:9000])
```

```python
p.predict(['söyle sevgilim hadi söyle hiç mi mutlu olmadık martıları sayarken hiç mi hayal kurmadık denizlere bakarken'])
```
