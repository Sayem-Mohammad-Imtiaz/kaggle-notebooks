```python
def missing_zero_values_table(df):
    mis_val = df.isnull().sum()
    mis_val_percent = round(df.isnull().mean().mul(100),2)
    mz_table = pd.concat([mis_val,mis_val_percent],axis=1)
    mz_table = mz_table.rename(columns = {df.index.name:'col_name',0:'Missing Values',1:'% of Total Values'})
    mz_table['Data_type']=df.dtypes
    mz_table=mz_table.sort_values('% of Total Values',ascending=False)
    print("Your selected dataframe has ** "+str(df.shape[1])+" ** columns and ** "+str(df.shape[0])+" ** Rows.\n"
                 "There are ** "+str(mz_table[mz_table.iloc[:,1] != 0].shape[0])+
                  " ** columns that have missing values. \n")
    return mz_table.reset_index()
```

```python
missing_zero_values_table(df)
```

# Visualizing the Data to Gain insights

```python
mostTarget = go.Bar(x=tweets_df.groupby('target')['Unnamed: 0'].count().nlargest(15).index,
             y=tweets_df.groupby('target')['Unnamed: 0'].count().nlargest(15))
iplot([mostTarget])
```

```python
stopwords = set(STOPWORDS)
totalTarget = ' '.join(tweets_df.target.astype(str))
wordCloud = WordCloud(width=1000,height=500,stopwords=stopwords).generate(totalTarget)
```

```python
fig = px.imshow(wordCloud, color_continuous_scale='gray')
fig.update_layout(coloraxis_showscale=False)
fig.update_xaxes(showticklabels=False)
fig.update_yaxes(showticklabels=False)
fig.show()
```
