**What is the percentage that team which First choose the Bat and win the match?**

```python
bat_and_won = matches[(matches['toss_decision'] == 'bat') & 
                      (matches['toss_winner'] == matches['winner'])]
print(f'Percentage : {(len(bat_and_won)/len(matches)*100)}')
```

```python
field_and_won = matches[(matches['toss_decision'] == 'field') & 
                        (matches['toss_winner'] == matches['winner'])]
print(f'Percentage : {(len(field_and_won)/len(matches)*100)}')
```

**So can conclude that most of the time chasing is simple in IPL matches. So if you choose the Filed first time then your chance of winning match get increases near about 50% **

```python
matches_winner = matches['toss_winner'] == matches['winner']
matches_winner = matches_winner.value_counts()
labels = ["Win The match",'Loss The Match']
```

***Change data attribute from object to date for analysis***

```python
temp_df = matches.drop_duplicates(subset=['season'],
                                  keep='last')[['season', 'winner']].reset_index(drop=True)
temp_df.sort_values(by='season')
```

```python
temp_df = pd.melt(matches, id_vars=['id'], value_vars=['umpire1', 'umpire2','umpire3'])
temp_series = temp_df.value.value_counts()[:10]
labels = np.array(temp_series.index)
ind = np.arange(len(labels))
width = 0.9
fig, ax = plt.subplots()
rects = ax.bar(ind, np.array(temp_series), width=width, color='g')
ax.set_xticks(ind+((width)/2.))
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Count")
ax.set_title("Top Umpires")
plt.show()
```

S. Ravi seems to be the most sought after umpire for IPL matches followed by Dharmasena. Others are fairly close to each other.

```python
total_run_of_players = deliveries.groupby(['batsman'])['total_runs'].count()
total_run_of_players = pd.DataFrame(total_run_of_players)
top_20_runners = total_run_of_players.sort_values(by='total_runs',ascending=False).head(20)
```

```python
top_20_runners.plot(kind='bar',
                   title = 'Top 20 Runners',
                   color=sns.color_palette('BuPu_r',100),
                   figsize=[20,10])
plt.show()
```

**Virat Kohli is the leading batsman according to our analysis followed closely by Raina**

```python
batsman_with_most_number_of_boundires = deliveries.groupby('batsman')['batsman_runs'].agg(lambda x : 
                    (x == 4).sum()).reset_index().sort_values(by='batsman_runs',
                    ascending=False).reset_index(drop=True)
top_20_players_with_most_boundires = batsman_with_most_number_of_boundires[:20]
```

```python
boundires = deliveries.groupby('batsman')['batsman_runs'].agg(lambda x : 
                    (x == 4).sum())
boundires = pd.DataFrame(boundires)
top_20_boundires_played = boundires.sort_values(by='batsman_runs',ascending=False).head(20)
top_20_boundires_played.plot(kind='bar',
                   title = 'Top 20 Played Boundries by Players',
                   color=sns.color_palette('Oranges',1),
                   figsize=[20,10])
plt.show()
```

Dhawan played more boundires than all other players

```python
boundires = deliveries.groupby('batsman')['batsman_runs'].agg(lambda x : 
                    (x == 6).sum())
boundires = pd.DataFrame(boundires)
top_20_sixers_played = boundires.sort_values(by='batsman_runs',ascending=False).head(20)
top_20_sixers_played.plot(kind='bar',
                   title = 'Batsman with most number of sixes',
                   color=sns.color_palette('autumn',2),
                   figsize=[20,10])
plt.show()
```

**Yeah, as we all know Gayle is famous for his sixes**

```python
top_10_dot_balls = deliveries.groupby(['batsman'])['wide_runs'].agg(lambda x : 
                    (x == 0).sum())
top_10_dot_balls = pd.DataFrame(top_10_dot_balls)
top_10_dot_balls = top_10_dot_balls.sort_values(by='wide_runs',ascending=False).head(20)
top_10_dot_balls.plot(kind='bar',
                   title = 'Bowlers with most number of dot bowles',
                   color=sns.color_palette('CMRmap',2),
                   figsize=[20,10])
plt.show()
```

```python
top_10_wide_bowls = deliveries.groupby(['bowler'])['wide_runs'].count()[:10]
top_10_wide_bowls = pd.DataFrame(top_10_wide_bowls)
top_10_wide_bowls = top_10_wide_bowls.sort_values(by='wide_runs',ascending=False).head(20)
top_10_wide_bowls.plot(kind='bar',
                   title = 'Bowlers with most number of wide bowles',
                   color=sns.color_palette('flag_r',2),
                   figsize=[20,10])
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='dismissal_kind', data=deliveries)
plt.xticks(rotation='vertical')
plt.show()
```

**Caught is the most common dismissal type in IPL followed by Bowled. There are very few instances of hit wicket as well. 'Obstructing the field' is one of the dismissal type as well in IPL.!**
