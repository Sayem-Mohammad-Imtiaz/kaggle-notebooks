```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (32, 14)
plt.rcParams["font.size"] = 21

df = pd.read_csv("/kaggle/input/wifi-of-the-access-point-where-i-live/wifipoints.csv")

tarr = [' 300 Mbps', ' 144 Mbps', ' 150 Mbps', ' 130 Mbps', ' 72 Mbps', ' 1733 Mbps', ' 54 Mbps', ' 450 Mbps', ' 216 Mbps', ' 270 Mbps', ' 867 Mbps', ' 65 Mbps', ' 433 Mbps', ' 3380 Mbps']
outs = [0.0]*len(tarr)

maximumspeeds = df['maximumspeeds']

for i in range(0, len(maximumspeeds)):
    outs[tarr.index(maximumspeeds[i])] += 1

xx = []
yy = []

for i in range(0, len(tarr)):
    tarr[i] = [tarr[i], outs[i]]
tarr.sort(key=lambda xx: xx[1], reverse=True)

for i in range(0, len(tarr)):
    xx.append(tarr[i][0])
    yy.append(tarr[i][1])
    
plt.plot(xx, yy, marker="o", linewidth=1.618)
plt.xticks(rotation=45)
plt.title("Number of points by speed.")
plt.xlabel("Maximum speeds")
plt.ylabel("Counter")
plt.grid(linestyle="--", linewidth=1.618)
plt.show()
```
