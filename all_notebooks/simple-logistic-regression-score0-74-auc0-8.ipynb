```python
import pandas as pd
```

```python
"""read csv"""
df = pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
```

```python
"""check missing observations"""
df.isnull().sum() #no missing data
```

```python
"""split data for sklearn"""
y = df['Outcome']
X = df.iloc[:,:-1] #use df except for the last column
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y = train_test_split(X,y,test_size=0.3, random_state = 42)
```

```python
"""run regression to check if the variables are correlated"""
import statsmodels.api as sm
logit_model=sm.Logit(y,X)
result=logit_model.fit()
print(result.summary())
```

```python
"""run logistic regression"""
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(train_X,train_y)
log_reg.score(train_X,train_y) #0.782
log_reg.score(test_X,test_y) #0.740
log_predicted = log_reg.predict(test_X)
```

```python
"""show predicted results"""
log_predicted
```

```python
"""get auc score"""
probs = log_reg.predict_proba(test_X)
preds = probs[:,1]
from sklearn.metrics import roc_auc_score
auc_score = roc_auc_score(test_y, preds) #score = 0.80
```
