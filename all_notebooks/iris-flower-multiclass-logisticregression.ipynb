## Exploring the data

```python
import pandas as pd # Import padas to read csv file
import matplotlib.pyplot as plt # Import to visualize the data
%matplotlib inline
```

```python
# Read data set from csv
df = pd.read_csv('../input/iris-flower-dataset/IRIS.csv')
df.head()
```

```python
df.info()
```

```python
# Visualize data set
import seaborn as sns
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(df.sepal_length, hue=df.species)
```

From thhis countplot we found out the range of different species wehwn it comes to the sepal length

<h2>Splitting data set<h2>

```python
X = df.drop('species', axis='columns')
X.head()
```

```python
y = df.species
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
y_train
```

## Creating logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=200) # Initialize the model
```

```python
model.fit(X_train, y_train) # Traing the model
```

```python
model.score(X_test, y_test) #Measuring the accuracy of the model
```

93 percent accuracy is a good accuracy

```python
y_predicted = model.predict(X_test) #Predicting output using train data
y_predicted
```

```python
y_test.head()
```

## Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
import seaborn as sns
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
