Next, lets make some changes in our columns. Start with columns that contains of datetime (DAY_OF_MONTH,DAY_OF_WEEK, DEP_TIME, ARR_TIME)

To analyze time of flight in this case, I need to make DEP_TIME and ARR_TIME in datetime format. I am gonna start to make new columns that contains date for each flight.

```python
#2019

date = []
dept_time =[]
arr_time = []

for i in df_19_temp['DAY_OF_MONTH']:
# for i, value in enumerate(df_temp['DAY_OF_MONTH']):
    if len(str(i)) ==1:
        date.append('2019-01-0{0}'.format(i))
    else:
        date.append('2019-01-{0}'.format(i))
        
df_19_temp['DATE'] =date



for i in df_19_temp['DEP_TIME']:
    if len(str(i)) == 5:
        dept_time.append(str(i)[0:3][:1] + ':' + str(i)[0:3][1:])
    elif len(str(i)) == 6:
        dept_time.append(str(i)[0:4][:2] + ':' + str(i)[0:4][2:])
    else :
        dept_time.append('a')

for i in df_19_temp['ARR_TIME']:
    if len(str(i)) == 5:
        arr_time.append(str(i)[0:3][:1] + ':' + str(i)[0:3][1:])
    elif len(str(i)) == 6:
        arr_time.append(str(i)[0:4][:2] + ':' + str(i)[0:4][2:])
    else :
        arr_time.append('a')

df_19_temp['DEP_TIME'] = dept_time
df_19_temp['ARR_TIME'] = arr_time
df_19_temp.drop( df_19_temp[ df_19_temp['DEP_TIME'] == 'a' ].index , inplace=True)
df_19_temp.drop( df_19_temp[ df_19_temp['ARR_TIME'] == 'a' ].index , inplace=True)


df_19_temp.DEP_TIME = df_19_temp.DEP_TIME.replace({"24:00": "23:59"})
df_19_temp.ARR_TIME = df_19_temp.ARR_TIME.replace({"24:00": "23:59"})

df_19_temp['DEP_TIME'] = df_19_temp['DATE'] + ' ' +df_19_temp['DEP_TIME']
df_19_temp['ARR_TIME'] = df_19_temp['DATE'] + ' ' +df_19_temp['ARR_TIME']



import datetime

df_19_temp['DEP_TIME'] = pd.to_datetime(df_19_temp['DEP_TIME'],format= '%Y-%m-%d %H:%M' )
df_19_temp['ARR_TIME'] = pd.to_datetime(df_19_temp['ARR_TIME'],format= '%Y-%m-%d %H:%M' )
```

```python
#2020

date = []
dept_time =[]
arr_time = []

for i in df_20_temp['DAY_OF_MONTH']:
# for i, value in enumerate(df_temp['DAY_OF_MONTH']):
    if len(str(i)) ==1:
        date.append('2020-01-0{0}'.format(i))
    else:
        date.append('2020-01-{0}'.format(i))
        
df_20_temp['DATE'] =date



for i in df_20_temp['DEP_TIME']:
    if len(str(i)) == 5:
        dept_time.append(str(i)[0:3][:1] + ':' + str(i)[0:3][1:])
    elif len(str(i)) == 6:
        dept_time.append(str(i)[0:4][:2] + ':' + str(i)[0:4][2:])
    else :
        dept_time.append('a')

for i in df_20_temp['ARR_TIME']:
    if len(str(i)) == 5:
        arr_time.append(str(i)[0:3][:1] + ':' + str(i)[0:3][1:])
    elif len(str(i)) == 6:
        arr_time.append(str(i)[0:4][:2] + ':' + str(i)[0:4][2:])
    else :
        arr_time.append('a')

df_20_temp['DEP_TIME'] = dept_time
df_20_temp['ARR_TIME'] = arr_time
df_20_temp.drop( df_20_temp[ df_20_temp['DEP_TIME'] == 'a' ].index , inplace=True)
df_20_temp.drop( df_20_temp[ df_20_temp['ARR_TIME'] == 'a' ].index , inplace=True)


df_20_temp.DEP_TIME = df_20_temp.DEP_TIME.replace({"24:00": "23:59"})
df_20_temp.ARR_TIME = df_20_temp.ARR_TIME.replace({"24:00": "23:59"})

df_20_temp['DEP_TIME'] = df_20_temp['DATE'] + ' ' +df_20_temp['DEP_TIME']
df_20_temp['ARR_TIME'] = df_20_temp['DATE'] + ' ' +df_20_temp['ARR_TIME']



import datetime

df_20_temp['DEP_TIME'] = pd.to_datetime(df_20_temp['DEP_TIME'],format= '%Y-%m-%d %H:%M' )
df_20_temp['ARR_TIME'] = pd.to_datetime(df_20_temp['ARR_TIME'],format= '%Y-%m-%d %H:%M' )
```

```python
from datetime import timedelta

#df_19_temp
for i,value in enumerate(df_19_temp['LONG_FLIGHT']):
    if df_19_temp['LONG_FLIGHT'][i].days <0:
        df_19_temp['LONG_FLIGHT'][i]= timedelta(days=0,seconds=df_19_temp['LONG_FLIGHT'][i].seconds).seconds/60
    else:
        df_19_temp['LONG_FLIGHT'][i] = timedelta(days=0,seconds=df_19_temp['LONG_FLIGHT'][i].seconds).seconds/60

#df_20_temp
for i,value in enumerate(df_20_temp['LONG_FLIGHT']):
    if df_20_temp['LONG_FLIGHT'][i].days <0:
        df_20_temp['LONG_FLIGHT'][i]= timedelta(days=0,seconds=df_20_temp['LONG_FLIGHT'][i].seconds).seconds/60
    else:
        df_20_temp['LONG_FLIGHT'][i] = timedelta(days=0,seconds=df_20_temp['LONG_FLIGHT'][i].seconds).seconds/60
```

```python
df_origin_depdel_2019 = pd.crosstab(df_flight[df_flight['YEAR']==2019]['ORIGIN'], 
                    df_flight[df_flight['YEAR']==2019]['DEP_DEL15'],normalize='index').unstack().reset_index().rename(columns={0: "RATE"})
df_origin_depdel_2020 = pd.crosstab(df_flight[df_flight['YEAR']==2020]['ORIGIN'], 
                df_flight[df_flight['YEAR']==2020]['DEP_DEL15'],normalize='index').unstack().reset_index().rename(columns={0: "RATE"})

df_origin_depdel = pd.merge(df_origin_depdel_2019, df_origin_depdel_2020, how='left', on=['ORIGIN','DEP_DEL15'])[['DEP_DEL15','ORIGIN','RATE_x','RATE_y']].rename(columns={"RATE_x": "RATE_2019", "RATE_y": "RATE_2020"})
df_origin_depdel['RATE_2019'] = df_origin_depdel['RATE_2019'].round(3)*100
df_origin_depdel['RATE_2020'] = df_origin_depdel['RATE_2020'].round(3)*100
df_origin_depdel['GROWTH'] = df_origin_depdel['RATE_2020'] - df_origin_depdel['RATE_2019']
df_origin_depdel['DEP_DEL15'] = df_origin_depdel['DEP_DEL15'].astype(int)
df_origin_depdel =df_origin_depdel[df_origin_depdel['DEP_DEL15']==0]
df_origin_depdel.head()
```

```python
df_scat_origin = pd.merge(df_origin_depdel, 
                             result_orig, 
                             how='left', 
                             on=['ORIGIN'])[['DEP_DEL15'
                                             ,'ORIGIN'
                                             ,'RATE_2019'
                                             ,'count_2019'
                                             ,'RATE_2020'
                                             ,'count_2020'
                                             ,'GROWTH_x'
                                             ,'GROWTH_y']].rename(columns={"count_2019": "TOTAL_FLIGHT_2019"
                                                                           ,"count_2020": "TOTAL_FLIGHT_2020"
                                                                           ,"RATE_2019" : "ONTIME_RATE_2019"
                                                                           ,"RATE_2020" : "ONTIME_RATE_2020"
                                                                           ,"GROWTH_x": "ONTIME_RATE_GROWTH"
                                                                           ,"GROWTH_y": "GROWTH_ORIGIN_AIRPORT"})
df_scat_origin = df_scat_origin[df_scat_origin['TOTAL_FLIGHT_2020'].notnull()].sort_values(by='TOTAL_FLIGHT_2020', ascending=False).reset_index(drop=True)
```
