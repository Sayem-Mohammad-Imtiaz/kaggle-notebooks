```python
sns.pairplot(data.drop('Id' ,axis=1), hue='Species', markers='o')
```

```python
fig, axes = plt.subplots(2,2,figsize=(16,12))
fig.suptitle('Sepal Petal Length Variations')
fig.tight_layout(pad=6.0)
sns.violinplot(ax=axes[0,0], data=data, y='Species', x='SepalLengthCm', inner='quartile')
sns.violinplot(ax=axes[0,1], data=data, y='Species', x='SepalWidthCm', inner='quartile')
sns.violinplot(ax=axes[1,0], data=data, y='Species', x='PetalLengthCm', inner='quartile')
sns.violinplot(ax=axes[1,1], data=data, y='Species', x='PetalWidthCm', inner='quartile')
```

```python
X = data.drop(['Id', 'Species'], axis=1)
y = data.Species
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_valid, y_train, y_valid = train_test_split(X,y,test_size=0.33, random_state=5) 
```

```python
X_valid.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

```python
kvals = list(range(1,20))
scores={}
for k in kvals:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    predictions = knn.predict(X_valid)
    score = metrics.accuracy_score(predictions, y_valid)
    scores[k]=score
```

```python
plt.plot(scores.keys(), scores.values())
```
