```python
import numpy as np
import pandas as pd
df = pd.read_csv("../input/acidbasereaction/AcidBaseReaction.csv")
df.head()
```

```python
df.columns = df.columns.str.strip()
df.describe()
```

```python
import numpy
from sklearn.metrics import r2_score

x = df['HCL (ml)']
y = df['NaCl (g)']


mymodel = numpy.poly1d(numpy.polyfit(x, y, 2))

print("R squared for HCL (reactant 1) vs. NaCl (product)")
print(r2_score(y, mymodel(x)))
```

```python
x = df['NaHCO3 (g)']
y = df['NaCl (g)']


mymodel = numpy.poly1d(numpy.polyfit(x, y, 2))

print("R squared for NaHCO3 (reactant 2) vs. NaCl (product)")
print(r2_score(y, mymodel(x)))
```
