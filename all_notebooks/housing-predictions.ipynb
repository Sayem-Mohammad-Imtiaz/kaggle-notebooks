```python
import pandas as pd
import numpy as np
my_model= pd.read_csv('../input/housing-prices-competition-for-kaggle-learn-users/train.csv')
display(my_model)
```

```python
for coluna in my_model:
    if my_model[coluna].isnull().sum() > 10000:
        my_model = my_model.drop(coluna, axis=1)
print(my_model.isnull().sum())
```

```python
my_model.fillna(0, inplace = True)
```

```python
#my_model = my_model.dropna()

print(my_model.shape)
print(my_model.isnull().sum())
```

```python
display(my_model.dtypes)
display(my_model.iloc[0])
```

```python
y = my_model.SalePrice.astype(np.float32, copy = False)
my_features = ['LotFrontage', 'LotArea','MoSold' ,'YrSold']
X = my_model[my_features].astype(np.float32, copy = False)

display(y.head())
```

```python

print(X.head())
```

```python
from sklearn.tree import DecisionTreeRegressor
my_data_model = DecisionTreeRegressor(random_state = 1)
my_data_model.fit(X,y)
```
