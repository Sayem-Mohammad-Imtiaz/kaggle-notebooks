```python
data = pd.read_csv('../input/expenditure-budget-of-ministries-and-departments/Expenditure.csv')
data
```

```python
data2 = data[data['Ministries/Departments']=='Cabinet']
data2
```

```python
data2.dtypes
```

```python
data2["Budget 2019-2020 Total"]= data2["Budget 2019-2020 Total"].replace("...", "0.00")
data2
```

```python
data3=data2
data3['Budget 2019-2020 Total'] = pd.to_numeric(data3['Budget 2019-2020 Total'])
data3.dtypes
```

```python
table = pd.pivot_table(data3, values='Budget 2019-2020 Total', index='Detailed Head of Expenditure',
                    columns='Head of Expenditure', fill_value=0)
table
```

```python
fig = plt.figure(figsize=(12,12))
r = sns.heatmap(table, linewidths=1, linecolor='white')
r.set_title("Heatmap of Flight Density from 1949 to 1961")
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=go.Heatmap(z=table.values, x=table.columns, y=table.index))
fig.show()
```
