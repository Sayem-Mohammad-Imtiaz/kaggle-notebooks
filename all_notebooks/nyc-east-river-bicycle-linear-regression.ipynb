```python
df = pd.read_csv("../input/nyc-east-river-bicycle-counts.csv")
df
```


Since the day and date show the same value, their data is redundant. We would ideally like only one of the columns

```python
#df.isnull().values.any()
type(df['Precipitation'])
df = df.replace({
    'T': 0.01,
    '0.47 (S)': '0.47'
})
df = df.astype({"Precipitation": float})
df.isnull().values.any()
df.drop(columns=['Date', 'Unnamed: 0'])
df['Avg_Temp'] = (df["High Temp (°F)"] + df["Low Temp (°F)"]) / 2
df = df.astype({"Avg_Temp": float})
type(df['Avg_Temp'])
df
```

No missing values found.
In the precipitaion column, what does T and (S) mean?

T: "T" stand for "Trace". This is a small amount of precipitation that will wet a raingage but is less than the 0.01 inch measuring limit. Replacing T by 0.01
(S): Not sure what it stands for. Changing 0.47(S) to 0.47 to create a float datatype for precipiation column.

```python
import numpy as np
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Fiting regression model
X1= np.array(df.drop(['Day','Date','Total','Unnamed: 0','Brooklyn Bridge','Manhattan Bridge','Williamsburg Bridge','Queensboro Bridge','Avg_Temp'],1))
y1=np.array(df['Brooklyn Bridge'])
X1_train, X1_test, y1_train, y1_test = train_test_split(X1,y1, test_size = 0.2)
clf1 = LinearRegression()
clf1 = clf1.fit(X1_train, y1_train)
accuracy1 =clf1.score(X1_test,y1_test)
print('Accuracy: \n', accuracy1)
y1_pred = clf1.predict(X1_test)
print('Coefficients: \n', clf1.coef_)
# The mean squared error
print("Mean squared error: %.2f"
      % mean_squared_error(y1_test, y1_pred))
# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % r2_score(y1_test, y1_pred))
```

```python
import statsmodels.api as sm
import seaborn as sns
X = df["High Temp (°F)"]
#X = df["Precipitation"]
y = df["Manhattan Bridge"]
X = sm.add_constant(X)
# Note the difference in argument order
model = sm.OLS(y, X).fit()
predictions = model.predict(X) # make the predictions by the model
# Print out the statistics
model.summary()
```

```python
#Plot the regression line
sns.regplot(x='High Temp (°F)', y='Manhattan Bridge', data=df)
```
