<h2 style ="background:#333399;color:white;border:0"><b>Eksik Veri Nedir?</b></h2>

İncelenen veri setindeki gözlemlerde eksiklik olması durumunu ifade etmektedir.

<a id="2"></a> <br>
<h2 style ="background:#333399;color:white;border:0"><b>Eksik Veri Türleri Nelerdir?</b></h2>

<b>Tümüyle Rastlantısal Kayıp:</b> Diğer değişkenlerden ya da yapısal bir problemden kaynaklanmayan tamamen rastgele oluşan gözlemler

<b>Rastlantısal Kayıp:</b> Diğer değişkenlere bağlı olarak oluşabilen eksiklik türü.

<b>Rastlantısal Olmayan Kayıp:</b> Göz ardı edilemeyecek olan ve yapısal problemler ile ortaya çıkan eksiklik türü

<h2 style ="background:#333399;color:white;border:0"><b>Eksik Verilerin Anlamlılığı</b></h2>

Eksik değerler bazı durumlarda bilgi ifade edebilirler. Örneğin bir maaş anketinde 5k altı maaş olan kişiler kasıtlı olarak maaş alanını boş bırakabilirler. Bu durumda ayrı bir kolon açılıp değerin daha önceden NA olup olmadıgının belirtilmesi gerekir. Örneğin Maas kolonunda NA değerler varsa, "maas-NA" adında bir kolon oluşturulur ve önceden değerleri olanlara F, değeri olmayanlara (NA)  T girilebilir. <h2 style ="background:#333399;color:white;border:0"><b>Eksik Veriyi Direk Silmenin Zararları</b></h2>

Eksik değerlere sahip gözlemlerin veri setinden direk çıkarılması ve rasallığının incelenmemesi, yapılacak olan istatistiksel çıkarımların, modelleme çalışmalarının güvenilirliğini düşürecektir.

<a id="9"></a> <br>
<h2 style ="color:#f3ca20;background:#000000;border:0;"><b>Eksik Veri Analizi</b></h2>

```python
# Hangi Değişkende Ne Kadar Eksik Veri Var ? 

df.isnull().sum()
```

```python
# Eksik Veri Tüm Veri Setinin Yüzde Kaçı?

total_cells = np.product(df.shape)
total_missing = df.isnull().sum().sum()
print("Toplam Eksik Veri: %",(total_missing/total_cells)*100)
```

```python
# Rassalığı Analiz Etmek İçin

msno.matrix(df);
```

```python
# Yüksek Korelasyonlar Eksik Değerlerin Aynı Satırda Olduğunu Doğrular.

msno.heatmap(df);
```

<h2 style ="color:#f3ca20;background:#000000;border:0;"><b>Eksik Verileri Doldurma</b></h2>

<h3><font style="color:#e52165"> Ortalama İle Doldurma </font></h3>

```python
# Runtime Değişkeni İle İşlem Yapalım.
# Runtime Değişkeni İçinde Toplam Kaç Tane Eksik Gözelem Var?

df.Runtime.isnull().sum()
```

```python
# Runtime Değişkeninin Eksik Değerlerini, 
# Runtime Değişkeninin Ortalaması (df.Runtime.mean()) ile
# dolduralım (fillna())

df.Runtime.fillna(df.Runtime.mean())
```

<h3><font style="color:#e52165">Belli Bir Değer İle Doldurma </font></h3>

```python
# Değerimizi Önceden Belirleyelim;

deger = 1
```

```python
# Belirlediğimiz değer ile runtime değişkeninin eksik verilerini dolduralım

df.Runtime.fillna(deger)
```

<h3><font style="color:#e52165">SimpleImputer İle doldurma</font></h3>

SimpleImputer ile eksik verileri doldurmak daha kolaydır. Fakat teoride yukarıda gösterdiğimiz yöntemler ile aynıdır.

```python
# Import işlemi yapmalıyız;

from sklearn.impute import SimpleImputer
```

```python
# missing_values: neleri doldurmak istediğimizi belirtiriz, np.nan eksik verileri doldurur
# strategy: hangi yöntemi kullanmak istediğimizi belirleriz, "mean" ortalamadır.

imputer = SimpleImputer(missing_values = np.nan,strategy="mean")
```

```python
# Yukarıdaki işlemde, yapmak istediğimiz işlemlerin detaylarını verdik.
# Şimdi ise bunu uygulamak istediğimiz değişken(ler)i belirtiyoruz.

imputer.fit_transform(df[["Runtime"]])
```

> Görmüş olduğunuz gibi, bir tane imputer nesnesi oluşturduğumuz zaman, bunu istediğimiz değişkenlere fit_transform edebiliyoruz.

<h2 style ="color:#f3ca20;background:#000000;border:0;"><b>Eksik Verileri Tahmine Dayalı Doldurma</b></h2>

<h3><font style="color:#e52165">KNN</font></h3>

```python
!pip install ycimpute  
# Gerekli Kütüphaneyi import ediyoruz.

from ycimpute.imputer import knnimput
```

```python
# Bu işlem için numeric datalar gerekli bu yüzden veri setimizden sadece numerik dataları çekiyoruz

df_numeric = df._get_numeric_data()
```

```python
# İleride işimize yaracağı için kolon isimlerini bir değişeken kaydettik
# df'yi array'e çevirdik cünkü algoritmada kullanacağız

var_names = list(df_numeric)
n_df = np.array(df_numeric)
```

```python
# k parametresi komşuluk sayısıdır. (KNN Algoritması ile alakalıdır. Araştırabilirsiniz.)
# Bu algorit
# Yukarıda, algoritma için array'e çevirdiğimiz "n_df" değişkenini kullanıyoruz.

dff = knnimput.KNN(k=4).complete(n_df)
```

```python
# array olan dff veri setimizi, DataFrame'e çeviriyoruz

dff = pd.DataFrame(dff,columns=var_names)
```
