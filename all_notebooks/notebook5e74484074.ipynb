```python
data = pd.read_csv('../input/loan-default-prediction/Default_Fin.csv')
data.head()
```

```python
data.drop('Index', inplace=True, axis=1)
```

```python
data.head()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
sns.countplot(data['Employed'])
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X, y = data.drop('Defaulted?', axis=1),data['Defaulted?'] 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

logreg.fit(X_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
def metrix(y_true, y_pred):
    print(accuracy_score(y_true, y_pred))
    print(confusion_matrix(y_true, y_pred))
    print(classification_report(y_true, y_pred))
    
```

```python
y_pred = logreg.predict(X_test)
metrix(y_test, y_pred)
```

```python
from imblearn.over_sampling import SMOTE 
```

```python
sm = SMOTE(random_state=42)
X, y = sm.fit_resample(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
metrix(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=200)

rfc.fit(X_train, y_train)
```

```python
X, y = data.drop('Defaulted?', axis=1),data['Defaulted?'] 
```

```python
y_pred = rfc.predict(X)
metrix(y, y_pred)
```

```python
params = {
    "n_estimators": [100, 200, 300],
    "criterion": ['gini', 'entropy'],
    "max_depth": [3, 5, 7, 9],
    "min_samples_split":[0.1, 0.3, 0.6, 0.8]
}
```

```python
from sklearn.model_selection import GridSearchCV

gs = GridSearchCV(estimator=rfc, param_grid=params, cv=5, verbose=1)

gs.fit(X_train, y_train)
```

```python
gs.best_score_
```

```python
gs.best_params_
```

```python
rfc_best = RandomForestClassifier(criterion= 'gini',
 max_depth= 9,
 min_samples_split= 0.1,
 n_estimators= 200)
```

```python
rfc_best.fit(X_train, y_train)
```

```python
X, y = data.drop('Defaulted?', axis=1), data['Defaulted?']
y_pred = rfc_best.predict(X)
metrix(y, y_pred)
```
