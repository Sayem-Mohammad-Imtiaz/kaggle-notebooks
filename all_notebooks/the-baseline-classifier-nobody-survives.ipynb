This is just the naive baseline classifier that I mentioned in my dataset task. It basically predicts that nobody survived the Estonia disaster and gets a pretty good score because only about 15% of the passengers actually survived. Predicting nobody survived means we're right 85% of the time, brilliant! This demonstrates why accuracy is generally not the preferred performance metric for classifiers, especially when we're dealing with an imbalanced dataset, which is the case for this ferry disaster.

```python
import numpy as np
import pandas as pd

pd.reset_option('^display.', silent=True)
pd.set_option('mode.chained_assignment', None)

# Load the full dataset
X = pd.read_csv('/kaggle/input/passenger-list-for-the-estonia-ferry-disaster/estonia-passenger-list.csv')

# Split X and y
y = X.Survived
X = X.drop(['Survived'], axis=1) 

# Make a naive classifer that says nobody survived
from sklearn.base import BaseEstimator
class NobodySurvivedClassifier(BaseEstimator):
    def fit(self, X, y=None):
        pass
    def predict(self, X):
        return np.zeros((len(X),1), dtype=bool)

nobody_survived_classifier = NobodySurvivedClassifier()
nobody_survived_classifier.fit(X[:1], y[:1])
y_pred = nobody_survived_classifier.predict(X[1:])

# Lets get an amazing accuracy score!
from sklearn.metrics import accuracy_score
print(f'Accuracy score for baseline classifier: {accuracy_score(y[1:], y_pred)}')
```
