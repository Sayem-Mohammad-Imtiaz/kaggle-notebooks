```python
#load libraries
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as ws
ws.filterwarnings("ignore")
```

```python
df = pd.read_csv("../input/ipl-2020-data/2008-2019 IPL data/2008-2019 IPL data/most_runs_average_strikerate_2008_to_2019.csv")
df.head()
```

```python
# EDA Starts
```

    <br>
<center> <p> &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; Top 10 run getters in IPL</p></center>
    <br>
   

```python
# Top 10 run scorer's in the the IPL
sns.set()
fig,ax = plt.subplots(figsize =(15,6))
top = df.sort_values(by="total_runs", ascending  =False).head(10)[["batsman", "total_runs"]]
sns.barplot(x = top["batsman"], y = top["total_runs"], palette="magma", ax=ax)
ax.set(title  ="Top 10 Scoring batsman in IPL")
plt.xticks(rotation =  30)
plt.show()
```

<b style ="color: #e11d74; font-size: 20px"> Inference </b>  
<p style ="font-size: 18px"> In top 10 high scoring batsman the top 3 Indian Batsman (May be after this IPL KL Rahul will be the member in the LIST )

<center> <p> &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; &nbsp; Top 10  highly dimissed  batsman in IPL</p></center>
    <br>
   

```python
# Top 10 run Dismissed batsman in the the IPL
sns.set()
fig,ax = plt.subplots(figsize =(15,6))
dismissed = df.sort_values(by="out", ascending  =False).head(10)[["batsman", "out"]]
sns.barplot(x = dismissed["batsman"], y = dismissed["out"], palette="inferno", ax=ax)
ax.set(title  ="Top 10 Dismissed batsman in IPL")
plt.xticks(rotation =  30)
plt.show()
```

<b style ="color: #e11d74; font-size: 20px"> Inference </b>  


<p style ="font-size: 18px"> 
    
All are Indians :)--------------  
    

The Indian players are extensively playing the IPL from the 2008 onwards hence the most dimissals were done to the indian players. Regarding players the RG Sharma (hitman) was featured in this list due to his presence in the DECCAN CHARGERS and also the all of them are either openers in the innings except Mahi (MSD) 

(Another  is most of then are right handers) 

Hence, we can infer that the rate of dismissals is more  in the openers. there can be certain reasons of it. like

```python
# Top 10 high Averaged players
sns.set()
fig,ax = plt.subplots(figsize =(15,6))
top_avg = df.sort_values(by="average", ascending  =False).head(10)[["batsman", "average", "total_runs"]].sort_values(by ="total_runs", ascending = False).head(5)
sns.barplot(x = top_avg["batsman"], y = top_avg["average"], palette="inferno", ax=ax)
ax.set(title  ="Top 10 high averaging batsman in IPL")
plt.xticks(rotation =  30)
plt.show()
```

<b style ="color: #e11d74; font-size: 20px"> Inference </b>  
<p style ="font-size: 18px"> Highest averaging batsman is MR. 360 [ABD].

    <br>
   


```python
temp = df.sort_values(by="numberofballs", ascending  =False).head(10)[["batsman","total_runs", "numberofballs"]]
temp["avg"] = temp["total_runs"] /  temp["numberofballs"]
```

```python
sns.set()
fig,ax = plt.subplots(figsize =(15,6))
sns.barplot(x = temp["batsman"], y = temp["avg"], palette="rainbow", ax=ax)
ax.set(title  ="Balls / Run Ratio")
plt.xticks(rotation =  30)
plt.show()
```

<b style ="color: #e11d74; font-size: 20px"> Inference </b>  
<p style ="font-size: 18px"> Chris Gayl Bears the Highest ball per runs ratio followed by David Warner and  "*THALA and CHINNA*"  (MSD and SK Raina)

```python
# Runs per out
runs_per_innings = df.sort_values(by="total_runs", ascending  =False).head(10)[["batsman", "total_runs", "out"]]
runs_per_innings["runs_per_inn"]=  runs_per_innings["total_runs"] / runs_per_innings["out"]
sns.set()
fig,ax = plt.subplots(figsize =(15,6))
sns.barplot(x=runs_per_innings["batsman"], y =runs_per_innings["runs_per_inn"], palette="Spectral_r")
ax.set(title= "Runs / innings by the top Scoring batsman")
plt.show()
```
