```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import plotly
import plotly.express as ex
import plotly.graph_objs as go
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Read data
data=pd.read_csv('/kaggle/input/forbes-billionaires-2021-30/forbes_billionaires_geo.csv')
```

```python
data.head()
```

```python
## Get the country wise net worth
data['NetWorth']=data['NetWorth'].astype(float)
```

```python
## group data by country on total net worth and count of billionaire
country_grp=data.groupby(by='Country').agg({'Name':'count','NetWorth':'sum'}).reset_index()
```

```python
country_grp.rename({'Name':'CountofName','NetWorth':'SumofNetworth'},axis=1,inplace=True)
country_grp['%CountofName']=round(country_grp['CountofName']/sum(country_grp['CountofName'])*100,2)
country_grp.sort_values(by='SumofNetworth',ascending=False,inplace=True)
country_grp.reset_index(drop=True,inplace=True)
country_grp['SumofNetworth']=round(country_grp['SumofNetworth'],2)
```

<br>
<h1 style = "font-size:40px; font-family:Garamond ; font-weight : normal; background-color: #f6f5f5 ; color : #fe346e; text-align: center; border-radius: 100px 100px;">Dual Axis plot : Total worth & Percentage count of billionaires By Country</h1>
<br>

```python
### Top 20 country wise total worth and % count of billionaires
from plotly.subplots import make_subplots
import plotly
import plotly.graph_objs as go
fig=go.Figure()
fig.add_trace(go.Bar(
    x=country_grp['Country'][:20],
    y=country_grp['SumofNetworth'][:20],
    name='Total Net worth',
    marker_color='aqua',
    text=country_grp['SumofNetworth'][:20],
    textposition='inside',
    yaxis='y1'
))
fig.add_trace(go.Scatter(
    x=country_grp['Country'][:20],
    y=country_grp['%CountofName'][:20],
    name='Percentage Billionaires',
    mode='markers+text',
    marker_color='red',
    marker_size=8,
    text=country_grp['%CountofName'][:20],
    textposition='top center',
    line=dict(color='#5D69B1',dash='dash'),
    yaxis='y2',
    texttemplate="%{text}%"

))
fig.update_layout(
    title="Top 20 Country : Forbes Billionaires Distribution and Networth By Country",
    xaxis_title="Country",
    yaxis_title="Total Worth",
    template='plotly_dark',
    font=dict(
        size=10,
        color="white",
        family="Courier"
    ),
    xaxis=dict(showgrid=True),
    yaxis=dict(showgrid=True),
    yaxis2=dict(showgrid=True,overlaying='y',side='right',title='% Billionaires'),
    legend=dict(
                        orientation='h',
                        yanchor="bottom",
                        y=-0.5,
                        xanchor="center",
                        x=0.5,
                        bgcolor="lightgrey",
                        bordercolor="black",
                        borderwidth=2,
                        font=dict(
                                family="Courier",
                                size=12,
                                color="black"
                            ))
)
fig.show()
```

```python
data['Age'].max()
```

```python
### Bucket the age's
data['Age_group']=np.where((data['Age']>=18) & (data['Age']<=28),'18-28yrs',
                          np.where((data['Age']>28) & (data['Age']<=38),'28-38yrs',
                                  np.where((data['Age']>38) & (data['Age']<=48),'38-48yrs',
                                          np.where((data['Age']>48) & (data['Age']<=58),'48-58yrs',
                                                  np.where((data['Age']>58) & (data['Age']<=68),'58-68yrs',
                                                          np.where(data['Age']>68,'>68yrs','NA'))))))
```

```python
data[data['Age_group']=='NA'].shape
```

<h1 style = "font-size:40px; font-family:Garamond ; font-weight : normal; background-color: #f6f5f5 ; color : #fe346e; text-align: center; border-radius: 100px 100px;">Plot for Age group</h1>
<br>

```python
### group data by age groups on total net worth and count of billionaires
age_grp=data.groupby(by='Age_group').agg({'Name':'count','NetWorth':'sum'}).reset_index()
```

```python
age_grp.rename({'Name':'CountofName','NetWorth':'SumofNetworth'},axis=1,inplace=True)
age_grp['%CountofName']=round(age_grp['CountofName']/sum(age_grp['CountofName'])*100,2)
age_grp.sort_values(by='SumofNetworth',ascending=False,inplace=True)
age_grp.reset_index(drop=True,inplace=True)
age_grp['SumofNetworth']=round(age_grp['SumofNetworth'],2)
```

<br>
<h1 style = "font-size:40px; font-family:Garamond ; font-weight : normal; background-color: #f6f5f5 ; color : #fe346e; text-align: center; border-radius: 100px 100px;">Pie chart - forbes Billionaire distribution and Net worth</h1>
<br>

```python
### Top 20 country wise total worth and % count of billionaires
from plotly.subplots import make_subplots
import plotly
import plotly.graph_objs as go
fig=go.Figure()
fig.add_trace(go.Pie(
    labels=age_grp[age_grp['Age_group']!='NA']['Age_group'],
    values=age_grp[age_grp['Age_group']!='NA']['SumofNetworth'],
    name='Total Net worth',
    text=age_grp[age_grp['Age_group']!='NA']['SumofNetworth']
))
fig.update_layout(
    title="Forbes Billionaires Distribution and Networth by Age Groups",
    template='plotly_dark',
    font=dict(
        size=10,
        color="yellow",
    ),
    plot_bgcolor='white',
    legend=dict(
                        orientation='h',
                        yanchor="bottom",
                        y=-0.5,
                        xanchor="center",
                        x=0.5,
                        bgcolor="lightgrey",
                        bordercolor="black",
                        borderwidth=2,
                        font=dict(
                                family="Courier",
                                size=12,
                                color="black"
                            ))
)
fig.show()
```

<h1 style = "font-size:40px; font-family:Garamond ; font-weight : normal; background-color: #f6f5f5 ; color : #fe346e; text-align: center; border-radius: 100px 100px;">Box PLot - Net worth by Status</h1>
<br>

```python
fig=go.Figure()
fig.add_trace(go.Box(x=data['Status'],
                     y=data['NetWorth'],
                     marker=dict(
                                color='limegreen',
                                outliercolor='red',
                                line=dict(
                                    outliercolor='red',
                                    outlierwidth=2)),
                     name="status",
                     jitter=0.5,
                     boxmean='sd',
                     boxpoints='suspectedoutliers'))
fig.update_xaxes(title_text="Status : Boxplot")
fig.update_layout(title={'text':"Boxplot :NetWorth Hue on Status",
                         'xanchor': 'center',
                         'yanchor': 'top',
                         'x':0.5,'y':0.97},
                      template='plotly_dark',
                      legend=dict(yanchor="bottom",
                                  y=0.5,
                                  xanchor="center",
                                  x=1.2),
                      barmode='overlay',
                      annotations=[dict(showarrow=True,
                                x=2,
                                y=72,
                                text=" Outlier",
                                xanchor="right",
                                xshift=0,
                                opacity=0.9,
                                font=dict(color="lightgoldenrodyellow",
                                          size=12),
                                arrowcolor="lightgoldenrodyellow",
                                arrowsize=5,
                                arrowwidth=0.5,
                                arrowhead=4)],
                     font=dict(
                                size=10,
                                color="yellow",
                                family="Carrier"
                            ),)
fig.update_traces(opacity=0.5)
fig.show();
```
