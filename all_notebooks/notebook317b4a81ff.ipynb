```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../input/us-inflation-data-updated-till-may-2021/US CPI.csv')
df.head()
```

```python
df['CPI'].describe()
```

```python
df['Yearmon'] = pd.to_datetime(df['Yearmon'], format='%Y%m')
df['Year'] = df['Yearmon'].dt.year
df['Month'] = df['Yearmon'].dt.month
df1 = df.drop(['Yearmon'], axis = 1)
df1.head()
```

```python
Year_CPI_Min = df1.groupby('Year')['CPI'].min().plot(kind = 'line', legend = True, figsize = (20,7),marker = 'v')
Year_CPI_Max = df1.groupby('Year')['CPI'].max().plot(kind = 'line', legend = True, figsize = (20,7),marker = 'o')
Year_CPI_avg = df1.groupby('Year')['CPI'].std().plot(kind = 'line', legend = True, figsize = (20,7),marker = 'D')
plt.title('Min, Max and STD of CPI by Year')
```
