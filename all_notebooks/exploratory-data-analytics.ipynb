```python
import matplotlib.pyplot as plt; plt.rcdefaults()

objects = data[data.Medal=="Gold"].Team.value_counts(ascending=False).index.tolist()
y_pos = np.arange(len(objects))
performance = data[data.Medal=="Gold"].Team.value_counts(ascending=False).tolist()

plt.bar(y_pos, performance, align='center', alpha=0.5, color="r")
plt.xticks(y_pos, objects)
plt.ylabel('Number of Gold medals won')
plt.xlabel("Teams")
plt.title('Performance')

plt.show()
```

```python
#21
df = pd.concat([data[data.City=="Beijing"] , data[data.City== "London"], data[data.City== "Sydney"], data[data.City== "Los Angeles"]])
df
```

```python
## data to plot
n_groups = 4

gold = df[df.Medal=="Gold"].City.value_counts(ascending=False).tolist()
silver = df[df.Medal=="Silver"].City.value_counts(ascending=False).tolist()
bronze = df[df.Medal=="Bronze"].City.value_counts(ascending=False).tolist()
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.20
opacity = 0.3

rects1 = plt.bar(index+ bar_width, gold, bar_width,
alpha=opacity,
color='yellow',
label='GOLD')

rects2 = plt.bar(index + 2*bar_width, silver, bar_width,
alpha=opacity,
color='grey',
label='SILVER')

rects3 = plt.bar(index + 3*bar_width, bronze, bar_width,
alpha=opacity,
color='brown',
label='BRONZE')

plt.xlabel('CITIES')
plt.ylabel('MEDALS WON')
plt.title('CITY-WISE PERFORMANCE')
plt.xticks(index + bar_width, ('Beijing', 'London', 'Sydney', 'Los Angeles'))
plt.legend()

plt.tight_layout()
plt.show()
```
