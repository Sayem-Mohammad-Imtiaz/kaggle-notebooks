```python
#Code by Gabriel Preda

def plot_count(feature, title, df, size=1):
    '''
    Plot count of classes / feature
    param: feature - the feature to analyze
    param: title - title to add to the graph
    param: df - dataframe from which we plot feature's classes distribution 
    param: size - default 1.
    '''
    f, ax = plt.subplots(1,1, figsize=(4*size,4))
    total = float(len(df))
    g = sns.countplot(df[feature], order = df[feature].value_counts().index[:30], palette='Set2')
    g.set_title("Number and percentage of {}".format(title))
    if(size > 2):
        plt.xticks(rotation=90, size=8)
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x()+p.get_width()/2.,
                height + 3,
                '{:1.2f}%'.format(100*height/total),
                ha="center") 
    plt.show()
```

```python
plot_count("Level of development", "Level of development", df,2)
```

```python
plot_count("Women Entrepreneurship Index", "Women Entrepreneurship Index", df,4)
```

```python
plot_count("Female Labor Force Participation Rate", "Female Labor Force Participation Rate", df,4)
```

```python
#Code by Gabriel Preda https://www.kaggle.com/gpreda/explore-the-rainforest-soundscape
def plot_feature_distribution(df, feature, feature2, title, kde_mode=False, hist_mode=True):
    f, ax = plt.subplots(1,1, figsize=(12,6))
    for item in list(df[feature2].unique()):
        d_df = df.loc[df[feature2]==item]
        try:
            sns.distplot(df[feature], kde=kde_mode, hist=hist_mode, label=item)
        except:
            pass
    plt.legend(labels=list(df[feature2].unique()), bbox_to_anchor=(1, 1), loc='upper right', ncol=2)
    plt.title(title)
    plt.show()
```

```python
plot_feature_distribution(df, 'Women Entrepreneurship Index','Female Labor Force Participation Rate', "Women Entrepreneurship & Female Labor Force")
```

```python
plot_feature_distribution(df, 'Entrepreneurship Index','Inflation rate', "Entrepreneurship & Inflation rate")
```
