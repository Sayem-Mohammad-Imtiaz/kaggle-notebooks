# Fit, Predict,Score

```python
# Pipeline

def fit_predict_score(model, X_y_train_test):
    X_train, X_test, y_train, y_test = X_y_train_test
    # extracts and flattens the series y_train which has series and values to values only
    model.fit(X_train, y_train.values.ravel())
    y_pred = model.predict(X_test)
    score = accuracy_score(y_test, y_pred)
    print('Accuracy Score on test data :',score)
    return 
```

# Normalization (MinMaxScaler)

```python
# Pipeline with MinMaxScaler()

def MMScale_fit_predict_score(model, X_y_train_test):
    X_train, X_test, y_train, y_test = X_y_train_test
    scaler = MinMaxScaler() 
    X_train_scaled = scaler.fit_transform(X_train) 
    # extracts and flattens the series y_train which has series and values to values only
    model.fit(X_train_scaled, y_train.values.ravel()) 
    X_test_scaled = scaler.transform(X_test) 
    y_pred = model.predict(X_test_scaled) 
    score = accuracy_score(y_test, y_pred)
    print('Accuracy Score on test data :',score)
    return 
```

#Optimization (Repeated Cross Validation k-fold)

```python
# Pipeline for CV k-fold

def rcv_kfold(model, X,y):
  
    X = X.astype('float32')
    y = y.astype('float32')
    # evaluate the model
    cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
    n_scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
    # report model performance
    print('Accuracy: %.3f (%.3f)' %(mean(n_scores), std(n_scores)))
    return 
```

```python
#Naive Bayes (Gaussian)(Parameter to optimise and range below)     ##
parameter = 'var_smoothing'                               ##
parameter_range = np.linspace(0.00000000001,0.00000002,8000)            ##
# Update Classifier and parameter  in 'clf'line
#view the optimum parameter by plotting parameter vs Accuracy scove and test score 
training_accuracy = []
test_accuracy = []
for variable in parameter_range:
    #instantiate classifier
    clf = GaussianNB(var_smoothing=variable) ##
    #fit classifier
    clf.fit(X_train, y_train.values.ravel())
    #record training set accuracy
    training_accuracy.append(clf.score(X_train, y_train))
    # record generalization accuracy or test accuracy
    test_accuracy.append(clf.score(X_test, y_test))
    
# basic chart plot
plt.plot(parameter_range, training_accuracy, label="training accuracy")
plt.plot(parameter_range, test_accuracy, label="test accuracy")
print()
plt.ylabel("Accuracy")
plt.xlabel(str(parameter))
plt.legend()
print()
```
