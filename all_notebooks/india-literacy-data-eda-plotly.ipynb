```python
import geopandas as gpd

fp = r'/kaggle/input/indiageofiles/india-polygon.shp'
map_df = gpd.read_file(fp)
map_df.rename(columns={'st_nm': 'State'},inplace=True)
map_df.head() #check the head of the file
```

```python
map_df['State'].nunique()
```

```python
data_merge = map_df.merge(literacy_data_sort, on='State', how='left')
data_merge.head()
```

```python
fig, ax = plt.subplots(1, figsize=(10, 10))
ax.axis('off')
ax.set_title('Statewise Literacy Rate', fontdict={'fontsize': '25', 'fontweight' : '10'})

# plot the figure
data_merge.plot(column='Literacy',cmap='YlOrRd', linewidth=0.8, ax=ax, edgecolor='0', legend=True,markersize=[39.739192, -104.990337]);
```

```python
df1['geometry'] = map_df['State'].map(state_map)
```

```python
# for each row fetch state values
for index, row in df1.iterrows():
    #print(row['State'])
    val = row['State']
     
print(val)
# if this state value exits in state_map fetch corresponding geometry value
```

```python
def get_geo(state_name):
    
    if state_name.strip() in state_map.keys():
        return state_map.get(state_name.strip())
    else:
        return 0
```

```python
df1
```

```python
# for each row fetch state values
for index, row in df1.iterrows():
    #print(row)
    
    print(row['State'])
    print("==")
    
    s =get_geo(row['State']) 
    #print(s)
    
    print("===============")
    df1['geometry'][index] = s
    
    
    #df2['geometry']= s
    #print(row)
   
    #val = row['State']
    
```
