```python
import pandas as pd
import numpy as np
data = pd.read_csv("../input/database.csv")
```

```python
date = data.loc[len(data)-1000:len(data),'Date']
time = data.loc[len(data)-1000:len(data),'Time']

year = []
for date_i in date:
    year.append(int(date_i.split('/',2)[2]))
year = pd.Series(year, name = 'yangli_y')
    
month = []
for date_i in date:
    month.append(int(date_i.split('/',2)[0]))
month = pd.Series(month, name = 'yangli_m')    

day = []
for date_i in date:
    day.append(int(date_i.split('/',2)[1]))  
day = pd.Series(day, name = 'yangli_d')    
    
hour = []
for time_i in time:
    hour.append(int(time_i.split(':',2)[0]))    
hour = pd.Series(hour, name = 'yangli_h')    
```

```python
mag = data.loc[len(data)-1000:len(data),'Magnitude']
mag = pd.Series(pd.Series.tolist(mag), name = 'magnitude') # is there any quick method?
```

```python
data_new = year.to_frame().join([month.to_frame(), day.to_frame(), hour.to_frame(), mag.to_frame()])
```
