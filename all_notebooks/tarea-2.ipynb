## A) Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.graph_objs as go
from plotly.offline import iplot
```

1. Gráfique las siguientes funciones con matplotlib: 
* logaritmo natural (de 1 a 10) (1 pts)
* exponencial (de -5 a 5) (1pts)
* tangente hiperbólico (de -10 a 10) (1 pts)

Todas las funciones deben estár en la misma línea

```python
# Logaritmo natural
def natural_logarithm(lower_limit,upper_limit):
    numbers_range = range(lower_limit,upper_limit)
    return np.log(numbers_range)
lower_limit=1
upper_limit=10
log=natural_logarithm(lower_limit,upper_limit)
plt.plot(log)
```

```python
# Exponencial
def exponential(lower_limit,upper_limit):
    numbers_range = range(lower_limit,upper_limit)
    return np.exp(numbers_range)
lower_limit=-5
upper_limit=5
exp=exponential(lower_limit,upper_limit)
plt.plot(exp)
```

```python
def hyperbolic_tangent(lower_limit,upper_limit):
    numbers_range = range(lower_limit,upper_limit)
    return np.tanh(numbers_range)
lower_limit=-10
upper_limit=10
hyperbolic=hyperbolic_tangent(lower_limit,upper_limit)
plt.plot(hyperbolic)
```

## B) Plotly

1. Genere un gráfico de líneas utilizando plotly. Utilice el dataset `àvocados` que se encuentra a continuación. Para el eje x utilice la fecha (Date), para el eje y el precio promedio (AveragePrice) y para el color de las líneas utilice la región (Region). ¡Solamente debe graficar los precios para Los Angeles y San Francisco! (3 pts)

El dataset original se puede encontrar en [Avocado Prices](https://www.kaggle.com/neuromusic/avocado-prices)

```python
avocados = pd.read_csv('../input/avocado-prices/avocado.csv', index_col=0)
avocados['Date'] = pd.to_datetime(avocados['Date'], format='%Y-%m-%d')
avocados[avocados["region"].isin(['LosAngeles','SanFrancisco'])]
avocados = avocados[:300]
trace1=go.Scatter(x=avocados.Date,y=avocados.AveragePrice,mode='lines')
data = [trace1]
fig = dict(data = data)
iplot(fig)
```

2. Genere un gráfico 3d utilizando para el eje x el volumen total (TotalVolume), para el eje y el total de bolsas (TotalBags), para el eje z el Precio promedio (AveragePrice) y la región (Region) para el color. ¡No debe graficar el total de US! (3 pts)

```python
avocados = pd.read_csv('../input/avocado-prices/avocado.csv', index_col=0)
avocados['Date'] = pd.to_datetime(avocados['Date'], format='%Y-%m-%d')
avocados[-avocados["region"].isin(["TotalUS"])]
avocados = avocados[:300]
fig = go.Figure(data=[go.Mesh3d(x=(avocados['Total Volume']),
                   y=(avocados['Total Bags']),
                   z=(avocados['AveragePrice']),
                   opacity=0.5,
                   color='rgba(244,22,100,0.6)'
                  )])
fig.show()
```
