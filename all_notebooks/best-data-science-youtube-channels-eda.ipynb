### **Declaring a function that convert links to clickable**

```python
def clickable(path):    
    # returns the final component of a url
    final_url = os.path.basename(path)
      
    # convert the url into link
    return f'<a href="{path}">{final_url}</a>'
```

### **YouTube channels which has subscribers above 1M**

```python
above_1M = data[data['Above_1m'] == 1]
```

**Only 4 channels has subscribers above 1M**

```python
plt.figure(figsize=(9,5))

plt.title("YT channels whose subscribers is above 1M\n", size = 15)
plots = sns.barplot(x = 'YouTube Channel Name',
                 y = 'Appx_Actual_Subscribers',
                 data = above_1M)

for bar in plots.patches:
    _ = plots.annotate(format(bar.get_height(), '.2f'), 
                       (bar.get_x() + bar.get_width() / 2, 
                        bar.get_height() - (bar.get_height()-3.1) / 2),
                        ha='center',
                        va='center',
                        size=15, xytext=(0, 0),
                        bbox=dict(boxstyle="round4, pad=0.6", fc="w", ec="k", lw=2),
                        textcoords='offset points')


plt.ylabel("Subscriber (in Million)")
plots.set_xticklabels(plots.get_xticklabels(),rotation = 30);
```

```python
above_1M[['YouTube Channel Name','YouTube_Channel_Link']].style.format({'YouTube_Channel_Link' : clickable})
```

##### **Adding column in dataframe by taking log of `Appx_Actual_Subscribers` because some large numbers are present in`Appx_Actual_Subscribers` which squeezes the lowest on.**

```python
data["log_Appx_Actual_Subscribers"] = np.log(data['Appx_Actual_Subscribers'])
```

**From this dataset, most of the datascience Youtube channels has subscribers between 200k and 1M, which can also be shown below graph**

```python
plt.rcParams["figure.figsize"] = [15, 20]


plt.title("Average subscribers of Data Science Channels on Youtube \n", size = 25);

sns.barplot(y = "YouTube Channel Name", x = "log_Appx_Actual_Subscribers",
            data = data, palette="cool_r")

plt.axvline(data['log_Appx_Actual_Subscribers'].mean(),
            linestyle = '--',
            color = 'k',
            linewidth = 2);
```
