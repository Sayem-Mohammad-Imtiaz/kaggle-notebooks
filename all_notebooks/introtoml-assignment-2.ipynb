**LABEL ENCODING**

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()

df['OP_UNIQUE_CARRIER']=label.fit_transform(df['OP_UNIQUE_CARRIER']).astype(str)
df['TAIL_NUM']=label.fit_transform(df['TAIL_NUM']).astype(str)
df['DEST']=label.fit_transform(df['DEST']).astype(str)
df['Wind']=label.fit_transform(df['Wind']).astype(str)
df['Condition']=label.fit_transform(df['Condition']).astype(str)
X=df.drop('TAXI_OUT',1)
Y=df['TAXI_OUT']
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.10)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)

m=np.empty(7,dtype=float)
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
l=LinearRegression()
l.fit(X_train,Y_train)
p=l.predict(X_test)
m[0]=mean_squared_error(Y_test,p)**0.5
print(m[0])
```

KNN Neighbors Model

```python
from sklearn.neighbors import KNeighborsRegressor
knn=KNeighborsRegressor(n_neighbors=200)
knn.fit(X_train,Y_train)
p1=knn.predict(X_test)
m[1]=mean_squared_error(Y_test,p1)**0.5
print(m[1])
```

Lasso Regression

```python
from sklearn.linear_model import Lasso
la=Lasso(alpha=0.05,normalize=True)
la.fit(X_train,Y_train)
p2=la.predict(X_test)
m[2]=mean_squared_error(Y_test,p2)**0.5
print(m[2])
```

Ridge Regression

```python
from sklearn.linear_model import Ridge
ridgereg=Ridge(alpha=0.05,normalize=True)
ridgereg.fit(X_train,Y_train)
p3=ridgereg.predict(X_test)
m[3]=mean_squared_error(Y_test,p3)**0.5
print(m[3])
```

Naive Bayes

```python
from sklearn.linear_model import BayesianRidge
b=BayesianRidge()
b.fit(X_train,Y_train)
p4=b.predict(X_test)
m[4]=mean_squared_error(Y_test,p4)**0.5
print(m[4])
```

SVR Model

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
s1=StandardScaler()
s2=StandardScaler()
x1=X_train.copy()
x2=X_test.copy()
x1=s1.fit_transform(x1)
x2=s2.fit_transform(x2)
s=SVR()
s.fit(x1,Y_train)
p5=s.predict(x2)
m[5]=mean_squared_error(Y_test,p5)**0.5
print(m[5])
```

Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
r=RandomForestRegressor()
r.fit(X_train,Y_train)
p6=r.predict(X_test)
m[6]=mean_squared_error(Y_test,p6)**0.5
print(m[6])
```
