```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt

data = pd.read_csv("../input/mnist-in-csv/mnist_train.csv")

x_train = data.drop('label',axis = 1)
x_train = x_train.to_numpy()
x_train = x_train.reshape(-1, 28, 28, 1)

y_train = data['label'].to_numpy()



test_data = pd.read_csv("../input/mnist-in-csv/mnist_test.csv")
x_test = test_data.drop('label',axis = 1)
x_test = x_test.to_numpy()
x_test = x_test.reshape(-1, 28, 28, 1)

y_test = test_data['label'].to_numpy()

#normalize data

x_train = keras.utils.normalize(x_train, axis = 1)
x_test = keras.utils.normalize(x_test, axis = 1)

print(y_train[0])
#print(x_train[0])
```

```python
plt.imshow(x_train[0], cmap =plt.cm.binary)
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Flatten
from keras.layers import Dense
model = Sequential()
model.add(Flatten())
model.add(Dense(128, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(10, activation = 'softmax'))
model.compile(optimizer ='adam' ,
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'])
model.fit(x_train, y_train, epochs = 10)
```

```python
val_loss, val_acc = model.evaluate(x_test, y_test)
print(val_loss, val_acc)
```

```python
model.save('test_mnist.model')
```

```python
new_model = keras.models.load_model('test_mnist.model')
prediction = new_model.predict(x_test)
print(prediction)
```
