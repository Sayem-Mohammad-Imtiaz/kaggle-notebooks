```python
from dataprep.eda import create_report
import pandas as pd

df = pd.read_csv("../input/los-angeles-crime-data-20102020/Crime_Data_from_2020_to_Present.csv")
create_report(df)
```

```python
import sweetviz as sv

# EDA using Sweetviz
sweet_report = sv.analyze(pd.read_csv("../input/los-angeles-crime-data-20102020/Crime_Data_from_2020_to_Present.csv"))

# Shohwing the results
sweet_report.show_notebook(w="100%", h="full")
```

```python
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
df_viz = AV.AutoViz('../input/los-angeles-crime-data-20102020/Crime_Data_from_2010_to_2019.csv')
```
