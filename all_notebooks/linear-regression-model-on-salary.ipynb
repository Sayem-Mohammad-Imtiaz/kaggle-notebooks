### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importing Dataset

```python
dataset = pd.read_csv('/kaggle/input/years-of-experience-and-salary-dataset/Salary_Data.csv')
X = dataset.iloc[:, 0:1].values
y = dataset.iloc[:, 1].values
```

```python
print(X)
```

```python
print(y)
```

### Splitting Data into Training and Test Sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
print(X_train)
```

```python
print(X_test)
```

### Applying Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
rg = LinearRegression()
rg.fit(X_train, y_train)
```

### Visualizing Data

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, rg.predict(X_train), color = 'red')
```

#### Comparing Test Set with Training Set

```python
rg.fit(X_test, y_test)
plt.scatter(X_test, y_test)
plt.plot(X_train, rg.predict(X_train), color = 'red')
plt.plot(X_test, rg.predict(X_test), color = 'cyan')
```
