```python
df.Embarked.fillna(df.Embarked.mode()[0], inplace = True) 
```

```python
df.Age.fillna(df.Age.mean(), inplace=True)
```

```python
df.Embarked = df.Embarked.replace({'Q':0,'S':1,'C':2})
```

```python
X = df[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked']]
y = df.Survived
```

```python
X.shape, y.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X,y)
```

```python
X.values[10]
```

```python
dft.Sex = dft.Sex.replace({'male':0, 'female':1})
dft.Embarked = dft.Embarked.replace({'Q':0,'S':1,'C':2})
df.Embarked.fillna(df.Embarked.mode()[0], inplace = True)
dft.Age.fillna(dft.Age.mean(), inplace=True)
dft.Fare.fillna(dft.Fare.mean(), inplace=True)
```

```python
X_test = dft[['Pclass', 'Sex', 'Age', 'Fare', 'Embarked']]
```

```python
X_test.shape
```

```python
result = model.predict(X_test)
```
