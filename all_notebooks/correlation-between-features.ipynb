```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
df.head()
```

```python
list(df.columns)
```

```python
def get_details(col):
    print("Missing values present? ",df[col].isnull().any())
    print("Unique values:",df[col].unique())
    
```

```python
for col in list(df.columns):
    print("For",col)
    get_details(col)
    print("\n")
    
```

```python
male_data=[]
fem_data=[]
for x in range(1000):
    if df.iloc[x]['gender']=='male':
        male_data.append(df.iloc[x]['reading score'])
    else:
        fem_data.append(df.iloc[x]['reading score'])
        
```

```python
data=[male_data,fem_data]
fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(data)
 
# show plot
plt.show()
```

```python
#race and math score
a=[]
b=[]
c=[]
d=[]
e=[]
for x in range(1000):
    if df.iloc[x]['race/ethnicity']=='group A':
        a.append(df.iloc[x]['math score'])
    elif df.iloc[x]['race/ethnicity']=='group B':
        b.append(df.iloc[x]['math score'])
    elif df.iloc[x]['race/ethnicity']=='group C':
        c.append(df.iloc[x]['math score'])
    elif df.iloc[x]['race/ethnicity']=='group D':
        d.append(df.iloc[x]['math score'])
    else: 
        e.append(df.iloc[x]['math score'])
```

```python
data=[a,b,c,d,e]
fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(data)
 
# show plot
plt.show()
```

```python
stan=[]
free=[]
for x in range(1000):
    if df.iloc[x]['lunch']=='standard':
        stan.append(df.iloc[x]['writing score'])
    else:
        free.append(df.iloc[x]['writing score'])
        
data=[stan,free]
fig = plt.figure(figsize =(10, 7))
 
# Creating axes instance
ax = fig.add_axes([0, 0, 1, 1])
 
# Creating plot
bp = ax.boxplot(data)
 
# show plot
plt.show()
```
