# importing libraries

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

# importing dataset

```python
datf=pd.read_csv('/kaggle/input/kidney-stone-data/kidney_stone_data.csv')
```

```python
datf.shape
```

```python
datf.head(7)
```

# checking for missing values 

```python
datf.isnull().sum()
```

```python
datf.info()
```

# making independent features matrix

```python
x=datf[['treatment','stone_size']]
y=datf[['success']]
```

# dummification of x

```python
x=pd.get_dummies(x)
```

```python
print(x)
```

# splitting into training and test sets

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

# trying different classification models

```python
from sklearn.neighbors import KNeighborsClassifier
kncls=KNeighborsClassifier(n_neighbors=3,metric='minkowski',p=2)
kncls.fit(x_train,y_train)
y_pred1=kncls.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
cm1=confusion_matrix(y_test,y_pred1)
print(cm1)
accuracy_score(y_test,y_pred1)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
def mean(lst):
    i=0
    avg=0
    s=0
    while(i<len(lst)):
        s=s+lst[i]
        i=i+1
    avg=s/len(lst)
    return avg
```

```python
acc=cross_val_score(estimator=kncls,X=x_train,y=y_train,cv=10)
acc1=mean(acc)
print("Accuracy from kneighbors classifier model : {} %".format(acc1*100))
```

```python
from sklearn.naive_bayes import GaussianNB
gncls=GaussianNB()
gncls.fit(x_train,y_train)
y_pred2=gncls.predict(x_test)
```

```python
cm2=confusion_matrix(y_test,y_pred2)
print(cm2)
accuracy_score(y_test,y_pred2)
```

```python
acc=cross_val_score(estimator=gncls,X=x_train,y=y_train,cv=10)
acc2=mean(acc)
print("Accuracy from naive bayes classifier model : {} %".format(acc1*100))
```

```python
from sklearn.ensemble import RandomForestClassifier
rcls=RandomForestClassifier(n_estimators=20,criterion='entropy',random_state=0)
rcls.fit(x_train,y_train)
y_pred3=rcls.predict(x_test)
```

```python
cm3=confusion_matrix(y_test,y_pred3)
print(cm3)
accuracy_score(y_test,y_pred3)
```

```python
acc=cross_val_score(estimator=rcls,X=x_train,y=y_train,cv=10)
acc3=mean(acc)
print("Accuracy from randomforest classifier model : {} %".format(acc3*100))
```
