```python
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
import os
import glob
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
############################ Let's pull in the data#######################

CovidVacc = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv')
```

```python
CovidVacc.info()
```

```python
###################### Convert date from object to datetime format########
CovidVacc['date'] = pd.to_datetime(CovidVacc['date'], format = '%Y-%m-%d')
```

```python
####################### What countries are involved ? ######################
print("The Countries in the dataframe CovidVacc are\n")
CovidVacc.country.unique()
```

There are repeated data of the country of UK----- one for United Kingdom and separately of Scotland, England, Wales 
and Northern Ireland as we see below.

```python
########################To show repeated data of UK, England, Scotland, Wales and Northern Ireland########################
print("Last 2 days data of UK total_vaccinations:\n", 
      CovidVacc[CovidVacc['country'] == 'United Kingdom']['total_vaccinations'].tail(2))
print("\nLast 2 days data of England total_vaccinations:\n", 
      CovidVacc[CovidVacc['country'] == 'England']['total_vaccinations'].tail(2))
print("\nLast 2 days data of Scotland total_vaccinations:\n", 
      CovidVacc[CovidVacc['country'] == 'Scotland']['total_vaccinations'].tail(2))
print("\nLast 2 days data of Wales total_vaccinations:\n", 
      CovidVacc[CovidVacc['country'] == 'Wales']['total_vaccinations'].tail(2))
print("\nLast 2 days data of Northern Ireland total_vaccinations:\n", 
      CovidVacc[CovidVacc['country'] == 'Northern Ireland']['total_vaccinations'].tail(2))
```

When we add up the data of England, Scotland, Wales and Northern Ireland we get the consolidated data of United Kingdom. Hence we delete the rows of data of England, Scotland, Wales and Northern Ireland.

```python
print("Number of data ",CovidVacc.shape)
```

```python
########################## Deleting the rows of data ######################

CovidVacc = CovidVacc[~CovidVacc['country'].isin(['England', 'Scotland', 'Wales', 'Northern Ireland'])]
```

```python
print("The number of rows of data ",CovidVacc.shape)
```

```python
####################### Country leading by Total Vaccinations #################

fig, ax = plt.subplots(ncols=1, nrows=2, dpi=100, figsize=(28,24))

LeadTotVacc = CovidVacc.groupby(['country'])['total_vaccinations'].max().sort_values(ascending = False).head(20)
LeadTotVacc.sort_values(ascending = True).plot(kind = 'barh',
                                               ax=ax[0],color='dodgerblue', edgecolor='darkgrey')
ax[0].tick_params(axis='y', labelsize=22)
ax[0].set_xticklabels([0,50,100,150,200, 250], fontsize=22)
ax[0].set_ylabel(" ")
for label in ax[0].yaxis.get_ticklabels():
            label.wrap=True

ax[0].set_xlabel("In Millions", fontsize = 26)
Ystart, Yend = ax[0].get_ylim()
ax[0].vlines(x=[50e6,100e6,150e6,200e6,250e6],
             ymin=Ystart, ymax=Yend,  color='gray', linewidth = 0.7)
ax[0].set_title("Total Vaccinations By Leading Countries till date", fontsize=28)

############ Country leading by Total number of people vaccinated####################


LeadTotPeopleVacc = CovidVacc.groupby(['country'])['people_vaccinated'].max().sort_values(ascending = False).head(20)
LeadTotPeopleVacc.sort_values(ascending = True)
sns.barplot(x=LeadTotPeopleVacc.values, y=LeadTotPeopleVacc.index,
            ax=ax[1], color= 'orange', edgecolor='darkgrey')
ax[1].tick_params(axis='y', labelsize=22)
ax[1].set_xticklabels([0,20,40,60,80,100,120,140], fontsize=22)
ax[1].set_ylabel(" ")
ax[1].set_xlabel("In Millions",fontsize = 26)
for label in ax[1].yaxis.get_ticklabels():
            label.wrap=True
Ystart, Yend = ax[1].get_ylim()
ax[1].vlines(x=[20e6,40e6,60e6,80e6,100e6, 120e6,140e6],
             ymin=Ystart, ymax=Yend,  color='gray', linewidth = 0.7)
ax[1].set_title("Countries Leading By Total Number of People Vaccinated till date", fontsize=28)

plt.show()
```

China is leading by Total Number of Vaccinations done till date, as we see from the upper graph, followed closely by United States. India, United Kingdom and Brazil along with China and United States forms the leading 5 countries in vaccinations. 
But China is missing from the lower graph that shows the countries leading by number of people vaccinated.  
That's beacuse the data for the same is not available for China as we can see down below.
From the data that is available we can say US, India and UK are top 3 countries in getting their population vaccinated, followed by Brazil and Germany.

```python
################# Country leading by Total number of people vaccinated###################

fig, ax = plt.subplots(ncols=1, nrows=2, dpi=100, figsize=(26,22))

LeadTotPeopleVacc = CovidVacc.groupby(['country'])['people_vaccinated'].max().sort_values(ascending = False).head(20)

sns.barplot(x=LeadTotPeopleVacc.values, y=LeadTotPeopleVacc.index, edgecolor='navy', ax=ax[0])
ax[0].tick_params(axis='y', labelsize=22)
ax[0].set_xticklabels([0,20,40,60,80,100,120,140], fontsize=22)
ax[0].set_ylabel(" ")
ax[0].set_xlabel("In Millions",fontsize = 26)
for label in ax[0].yaxis.get_ticklabels():
            label.wrap=True
Ystart, Yend = ax[0].get_ylim()
ax[0].vlines(x=[20e6,40e6,60e6,80e6,100e6, 120e6,140e6], ymin=Ystart, ymax=Yend,  color='gray', linewidth = 0.7)
ax[0].set_title("Countries Leading By Total Number of People Vaccinated till date", fontsize=28)


#############################################Total number of people FULLY vaccinated####################################

LeadFulPeopleVacc = CovidVacc.groupby(['country'])['people_fully_vaccinated'].max().sort_values(ascending = False).head(20)

sns.barplot(x=LeadFulPeopleVacc.values, y=LeadFulPeopleVacc.index, edgecolor='navy', ax=ax[1])
ax[1].tick_params(axis='y', labelsize=22)
ax[1].set_xticklabels([0,20,40,60,80,100], fontsize=22)
ax[1].set_ylabel(" ")
#for label in ax[1].yaxis.get_ticklabels():
            #label.wrap=True

ax[1].set_xlabel("In Millions", fontsize = 26)
Ystart, Yend = ax[1].get_ylim()
ax[1].vlines(x=[20e6,40e6,60e6,80e6,100e6], ymin=Ystart, ymax=Yend,  color='gray', linewidth = 0.7)
ax[1].set_title("Countries Leading By Total Number of People FULLY Vaccinated till date", fontsize=28)


plt.show()
```

Without the data of China, we see that the top 5 countries leading in vaccinating its population are US, India, UK, Brazil and Germany. US leads in the number of people that received the entire set of immunization according to the immunization scheme. Below we explore the trend of Daily Vaccination of top 5 countries - US, India, UK, Brazil and Germany

```python
############## Daily vaccination Trend of Top 5 Countries leading in Total Number of People Vaccinated####################
 
print("Looking at the Daily Vaccination Trend of ", *LeadTotPeopleVacc.index[0:5], sep = ", ")
col = ['teal', 'green', 'gold', 'tomato', 'purple']

fig, ax = plt.subplots(ncols=1, nrows=5, dpi=100, figsize=(20,18))

for i in range(len(LeadTotPeopleVacc.index[0:5])):
    DailyVacc = CovidVacc[CovidVacc['country'] == LeadTotPeopleVacc.index[i]]
    sns.lineplot(data = DailyVacc, x='date', y='daily_vaccinations', linewidth=2.5, ax=ax[i],
                 label=LeadTotPeopleVacc.index[i], color=col[i])
    
    ax[i].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
    ax[i].xaxis.set_major_formatter(DateFormatter('%b%y'))
    ax[i].legend(prop={'size': 20}, loc='upper left')
    #Xstart, Xend = ax[i].get_xlim()
    #Ystep = round((Yend-Ystart)/5, 2)
    #ax[i].hlines(y=[2*Ystep, 4*Ystep], xmin=Xstart, xmax=Xend,  color='gray', linewidth = 0.7)
    ax[i].set_xlabel(' ')
    ax[i].set_ylabel(' ')
    ax[i].set_yticks([])
    ax[i].tick_params(axis='x', labelsize=14)
    
    
plt.show()
```

From the above graph we can see that United Kingdom has started an aggressive Vaccination drive as compared to other countries,
closely followed by India and United States

```python
################# Total number of people vaccinated per hundred #####################
############# Ratio (in percent) between population immunized and ###################
##################total population up to the date in the country #####################
####################### i.e % of population vaccinated ##############################

PopVaccinated = CovidVacc.groupby(['country'])['people_vaccinated_per_hundred'].max().sort_values(ascending = False).head(10)

PopFullyVaccinated = CovidVacc.groupby(['country'])['people_fully_vaccinated_per_hundred'].max().sort_values(ascending = False).head(10)

fig, ax = plt.subplots(ncols=1, nrows=1, dpi=100, figsize=(16,6))

sns.barplot(x=PopVaccinated.values, y=PopVaccinated.index, color='lightgreen',
            edgecolor='darkgrey', ax=ax)
for p in ax.patches:
    width = p.get_width()
    plt.text(p.get_width()-6, p.get_y()+0.50*p.get_height(),
            '{:1.2f}%'.format(width), fontsize=14,
             ha='center', va='center', color='black', fontweight='bold')
ax.set_ylabel('% of Population Vaccinated', fontsize = 16)
ax.tick_params(axis='y', labelsize=18)
ax.tick_params(axis='x', labelsize=14)

fig, ax = plt.subplots(ncols=1, nrows=1, dpi=100, figsize=(16,6))
sns.barplot(x=PopFullyVaccinated.values, y=PopFullyVaccinated.index,
            color='tan', edgecolor='darkgrey', ax=ax)
for p in ax.patches:
    width = p.get_width()
    plt.text(p.get_width()-6, p.get_y()+0.50*p.get_height(),
            '{:1.2f}%'.format(width), fontsize=14,
             ha='center', va='center', color='black', fontweight='bold')
ax.set_ylabel('% of Population Fully Vaccinated', fontsize = 16)
ax.tick_params(axis='both', labelsize=16)
```

Among the countries with the highest number of people vaccinated, Israel is an interesting case study. More than 62% of the population are vaccinated and 58% have completed the vaccination program.  
Let's look at the vaccination graph of Israel and compare it with India and United States.

Studying Israel's Vaccination against India and US

```python
################### Studying Israel Vaccination against India and US ###############

IsraelVacc = CovidVacc[CovidVacc['country'] == 'Israel']
IndiaVacc = CovidVacc[CovidVacc['country'] == 'India']
USVacc = CovidVacc[CovidVacc['country'] == 'United States']

fig, ax = plt.subplots(nrows=3, ncols=3, dpi=100, figsize=(22,18))

########################### Daily Vaccination ####################################

sns.lineplot(data = IsraelVacc, x='date', y='total_vaccinations',
             ax=ax[0,0], color = 'teal', linewidth=2)
sns.lineplot(data = IndiaVacc, x='date', y='total_vaccinations',
             ax=ax[0,1], color = 'teal', linewidth=2)
sns.lineplot(data = USVacc, x='date', y='total_vaccinations',
             ax=ax[0,2], color = 'teal', linewidth=2)
ax[0,0].set_ylabel("TOTAL VACCINATIONS", fontsize = 16)
ax[0,0].set_xlabel(" ")
ax[0,0].set_title("ISRAEL", fontsize=22)
ax[0,1].set_title("INDIA", fontsize=22)
ax[0,2].set_title("US", fontsize=22)
ax[0,0].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax[0,0].xaxis.set_major_formatter(DateFormatter('%b%y'))
ax[0,1].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax[0,1].xaxis.set_major_formatter(DateFormatter('%b%y'))
ax[0,2].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax[0,2].xaxis.set_major_formatter(DateFormatter('%b%y'))

############################## Daily vaccinations #############################
sns.lineplot(data = IsraelVacc, x='date', y='daily_vaccinations',
             ax=ax[1,0], color = 'lime', linewidth=2)
sns.lineplot(data = IndiaVacc, x='date', y='daily_vaccinations',
             ax=ax[1,1], color = 'lime', linewidth=2)
sns.lineplot(data = USVacc, x='date', y='daily_vaccinations',
             ax=ax[1,2], color = 'lime', linewidth=2)
ax[1,0].set_ylabel("DAILY VACCINATIONS", fontsize = 16)

ax[1,0].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax[1,0].xaxis.set_major_formatter(DateFormatter('%b%y'))
ax[1,1].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax[1,1].xaxis.set_major_formatter(DateFormatter('%b%y'))
ax[1,2].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax[1,2].xaxis.set_major_formatter(DateFormatter('%b%y'))

############################## % of Population Vaccinated #######################
sns.lineplot(data = IsraelVacc, x='date', y='people_vaccinated_per_hundred',
             ax=ax[2,0], color = 'coral', linewidth=2)
sns.lineplot(data = IndiaVacc, x='date', y='people_vaccinated_per_hundred',
             ax=ax[2,1], color = 'coral', linewidth=2)
sns.lineplot(data = USVacc, x='date', y='people_vaccinated_per_hundred',
             ax=ax[2,2], color = 'coral', linewidth=2)
ax[2,0].set_ylabel("% OF POPULATION VACCINATED", fontsize = 16)

ax[2,0].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax[2,0].xaxis.set_major_formatter(DateFormatter('%b%y'))
ax[2,1].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax[2,1].xaxis.set_major_formatter(DateFormatter('%b%y'))
ax[2,2].xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax[2,2].xaxis.set_major_formatter(DateFormatter('%b%y'))

for i in range(3):
    for j in range(3):
        ax[i,j].tick_params(axis='both', labelsize=16)
        ax[i,j].set_xlabel(" ")

for i in range(3):
    for j in range(2):
        ax[i,j+1].set_ylabel(" ") 
    
```

 The "TOTAL VACCINATIONS" curve is steepest for Israel as compared to India and US. That is because Israel had started with an
aggressive vaccination program and also have managed to boost up vaccination further as is shown by "DAILY VACCINATIONS" curve. 
The war-footing vaccination drive of Israel has resulted in vaccinating more than half of their population in a very short 
time as is visible from "% OF POPULATION VACCINATED" graph.

```python
VaccMan.info()
```

```python
############################### What countries are there in data? ####################################
print("The Countries within the country_vaccinations_by_manufacturer file are\n")
list(VaccMan.location.unique())
```

```python
##################### Who are Vaccine Manufacturers? #########################################
print("Who are the Vaccine manufacturers ?\n")
list(VaccMan.vaccine.unique())
```

```python
Vacc = VaccMan.groupby('vaccine').sum().sort_values(by=['total_vaccinations'], ascending=False)

fig, ax = plt.subplots(nrows=1, ncols=1, dpi=100, figsize = (10,6))

sns.barplot(x=Vacc.index, y=Vacc['total_vaccinations'], ax=ax)
ax.set_xlabel("Vaccine Manufacturers", fontsize = 16)
ax.set_ylabel(" ")

plt.show()
```

Pfizer/BioNTech vaccine is most widely used followed by Moderna. Sinovac, Oxford/AstraZeneca and Johnson&Johnson vaccines are less used vaccines. From the available data of 10 countries, we explore the vaccination data w.r.t manufacturers of 3 countries --- United States, France and Germany.

```python
####################### Convert date from object to datetime format#################
VaccMan['date'] = pd.to_datetime(VaccMan['date'], format = '%Y-%m-%d')
VaccMan.info()
```

```python
######################################################################################################################

print("Some Infomation on Vaccine Manufacturers by Countries \n")
USVaccByMan = VaccMan[VaccMan['location'] == 'United States'].groupby('vaccine').sum().sort_values(by=['vaccine'],
                                                                                                   ascending=False)
FranceVaccByMan = VaccMan[VaccMan['location'] == 'France'].groupby('vaccine').sum().sort_values(by=['vaccine'],
                                                                                                   ascending=False)
GermanyVaccByMan = VaccMan[VaccMan['location'] == 'Germany'].groupby('vaccine').sum().sort_values(by=['vaccine'],
                                                                                                   ascending=False)

fig, ax = plt.subplots(nrows=1, ncols=3, dpi=100, figsize=(24,8))

sns.barplot(x=USVaccByMan.index, y=USVaccByMan['total_vaccinations'].values, ax=ax[0])
ax[0].set_title("UNITED STATES", fontsize=22)

sns.barplot(x=FranceVaccByMan.index, y=FranceVaccByMan['total_vaccinations'].values, ax=ax[1])
ax[1].set_title("FRANCE", fontsize=22)
sns.barplot(x=GermanyVaccByMan.index, y=GermanyVaccByMan['total_vaccinations'].values, ax=ax[2])
ax[2].set_title("GERMANY", fontsize=22)

for i in range(3):
    ax[i].tick_params(axis='both', labelsize=16)
    ax[i].set_xlabel(" ")
    for label in ax[i].xaxis.get_ticklabels():
            label.set_rotation(60)


plt.show()
```

From the data available of country vaccinations w.r.t Covid Vaccine manufacturers, we see that Pfizer/BioNTech is the mostly 
used vaccine in United States, France and Germany. The UK manufactured Oxford/AstraZeneca, that is primarily used in India,
are used in less quantity in France and Germany but not in United States.
