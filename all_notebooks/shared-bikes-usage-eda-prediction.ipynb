```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
plt.style.use("bmh")
warnings.simplefilter("ignore")
```

```python
df = pd.read_csv("/kaggle/input/shared-bikes-demand-prediction/day.csv")
```

```python
df.head()
```

```python
df.drop(columns=["instant"], inplace=True)
```

```python
df.isna().sum()
```

```python
df["dteday"] = pd.to_datetime(df["dteday"])
```

```python
columns = ["temp", "atemp", "hum", "windspeed"]
fig, axs = plt.subplots(len(columns), 2, figsize=(25,15))
for i, column in enumerate(columns):
    df.plot(x=column, y="cnt", ax=axs[i,0], kind="scatter")
    df.plot(y=column, ax=axs[i,1], kind="kde") 
```

```python
temp_cut = df.groupby(pd.cut(df["temp"], 20)).sum()
ax = temp_cut.plot(y="cnt", kind="bar", figsize=(12,6))
ax.set_xticklabels(np.round((temp_cut.index.categories.right.values+temp_cut.index.categories.left.values)/2, 2));
```

```python
hum_cut = df.groupby(pd.cut(df["hum"], 20)).sum()
ax = hum_cut.plot(y="cnt", kind="bar", figsize=(12,6))
ax.set_xticklabels(np.round((hum_cut.index.categories.right.values+hum_cut.index.categories.left.values)/2, 2));
```

```python
wind_cut = df.groupby(pd.cut(df["windspeed"], 20)).sum()
ax = wind_cut.plot(y="cnt", kind="bar", figsize=(12,6))
ax.set_xticklabels(np.round((wind_cut.index.categories.right.values+wind_cut.index.categories.left.values)/2, 2));
```

```python
columns = ["casual", "registered"]
fig, axs = plt.subplots(len(columns), 1, figsize=(12,10))
for i, column in enumerate(columns):
    df.groupby("weekday").sum().plot(y=column, kind="bar", ax=axs[i])
```
