```python
# Importing Pandas As pd
import pandas as pd
# Importing Seaborn
import seaborn as sns
# Importing matplotlib
import matplotlib.pyplot as plt
# import KMeans from sklearn.cluster
from sklearn.cluster import KMeans
# Importing Nampy Array
import numpy as np
# Reading Data Frame
df =  pd.read_csv("../input/irisdataunsupervised/iris-data-unsupervised.csv")
df.head(6)
```

```python
# Reading The Info About The DataSet
df.info()
```

```python
# Reading class
df['Species-code'].value_counts()
```

```python
# Reading Data Clastering 
sns.scatterplot(x = "PetalLengthCm", y = "PetalWidthCm", hue="Species-code", style="Species-code", data=df)
# sizing the plot
sns.set(rc={'figure.figsize':(20,12)})
```

```python
# Clustering Data By KMeans Using ELBOW Method
distortions = []
K = range(1,10)
for k in K:
  kmeansModel = KMeans(n_clusters=k)
  kmeansModel.fit(df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']])
  distortions.append(kmeansModel.inertia_)

plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('distortion')
plt.title('Elbow Method')
plt.show()
```

```python
# Mentioning The Feature by the model will define Species
features = df[['PetalLengthCm', 'PetalWidthCm']]

model = KMeans(n_clusters=3, init='k-means++')
model.fit(features)

print(model.labels_)
```

```python
# Showing the calculated result in dataset
xx = model.labels_.tolist()
df['labels'] = xx
df
```

```python
# Showing The New result in scatterplot
sns.scatterplot(df['PetalLengthCm'], df['PetalWidthCm'], c=df['labels'])
```

```python
# Taking a Data to train my model
train_data= pd.read_csv("../input/irisdataunsupervised/traning.csv")
train_data
```

```python
# Running the model in traindata
e = model.predict(train_data)
e
```

```python
# Showing the calculated result in train dataset
train_data['Species'] = e
train_data
# replacing the number with Species name
train_data['Species'].replace(to_replace = 1, value = 'Iris-Setosa', inplace=True)
train_data['Species'].replace(to_replace = 2, value = 'Iris-Virginica', inplace=True)
train_data['Species'].replace(to_replace = 0, value = 'Iris-Versicolor', inplace=True)
train_data
```
