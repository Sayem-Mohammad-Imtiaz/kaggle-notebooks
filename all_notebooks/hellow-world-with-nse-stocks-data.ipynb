```python
data = pd.read_csv('../input/nse-stocks-data/FINAL_FROM_DF.csv')
```

```python
data['TRADE_DATE']= pd.to_datetime(data.TIMESTAMP,dayfirst=True)
data['DAILY_ABS_PCNT_CHANGE'] = (data.CLOSE-data.OPEN)*100/data.OPEN
```

```python
bank_stocks = data[data['SYMBOL'].str.contains("BANK")]
```

```python
bank_stocks.sort_index(inplace=False,ascending=True)
bank_stocks
```

```python
bank_stocks.SYMBOL.unique()
```

```python
bank_stocks.SYMBOL.shape
```

```python
stbdibank=bank_stocks.loc[bank_stocks['SYMBOL']=='SYNDIBANK','CLOSE']
stbdibank.sort_index(inplace=False,ascending=True)
```

```python
np.sort(stbdibank)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
plt.plot(stbdibank,linestyle='--',marker='o')
plt.xticks(np.arange(1,len(stbdibank),len(stbdibank)/15))
plt.show()
```

```python
bank_stocks_pivot = bank_stocks.pivot_table(values='CLOSE',columns='SYMBOL',index='TIMESTAMP')
bank_stocks_pivot.head()
```

```python
plt.figure(figsize=(30,20))
sns.heatmap(bank_stocks_pivot.corr(method='pearson'),annot=True,linewidth=0.9)
```

```python
bank_stocks_start=bank_stocks[bank_stocks.index==min(bank_stocks.index)]
bank_stocks_start.set_index('SYMBOL',inplace=True)

bank_stocks_end=bank_stocks[bank_stocks.index==max(bank_stocks.index)]
bank_stocks_end.set_index('SYMBOL',inplace=True)
bank_stocks_startend=pd.merge(bank_stocks_start['CLOSE'],bank_stocks_end['CLOSE'],on='SYMBOL')
bank_stocks_startend['GROWTH']=bank_stocks_startend['CLOSE_y'].div(bank_stocks_startend['CLOSE_x'])
bank_stocks_startend.columns = ['START', 'END','GROWTH']
```

```python
bank_stocks_startend.sort_values(by=['GROWTH'],ascending = False)
```
