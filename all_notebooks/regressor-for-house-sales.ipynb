```python
import pandas as pd
df = pd.read_csv('/kaggle/input/housesalesprediction/kc_house_data.csv')
df.head()
```

```python
X = df.drop(['price', 'id', 'date', 'zipcode', 'lat', 'long'  ], axis=1)
y = df['price']
```

```python
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(n_neighbors=5)

model.fit(X, y)
```

```python
model.score(X, y)
```

```python
y_pred = model.predict(X)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y, y_pred)
```

```python
model.predict([[3, 1, 1190, 5700, 1, 0, 0, 3, 7, 1190, 100, 1960, 0, 1370, 5700 ]])
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, shuffle=True)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor

model = KNeighborsRegressor(n_neighbors=5)

model.fit(X_train,y_train)

print('Train:', model.score(X_train, y_train))
print('Test:', model.score(X_test, y_test))
```

```python
from sklearn.metrics import mean_squared_error
y_pred_test = model.predict(X_test)
mean_squared_error(y_test, y_pred_test)
```

```python
from sklearn.model_selection import cross_validate

results = []
for k in [1,3,5,7,9,10,12,15,20]:
    model = KNeighborsRegressor(n_neighbors=k)
    cv_results = cross_validate(model, X_train, y_train, cv=3)
    results.append(sum(cv_results['test_score'])/3)
```

```python
results
```

```python
model = KNeighborsRegressor(n_neighbors=9)

model.fit(X_train,y_train)

print('Train:', model.score(X_train, y_train))
print('Test:', model.score(X_test, y_test))
```

# The same data that were for testing with k = 5. The model has become more accurate.
