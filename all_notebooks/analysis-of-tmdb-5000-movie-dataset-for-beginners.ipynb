# 2. PYTHON DATA SCIENCE TOOLBOX

**- USER DEFINED FUNCTION**

```python
def tuble_ex():
    t = (1,2,3,4)
    return t
a,b,c,d = tuble_ex()
print(a,b,c,d)
```

**- SCOPE**

```python
# Scope - Global/Local Scope
x = 5
def f():
    x = 7
    return x
print(x)
print(f())    
```

```python
# No local scope
x = 8
def f():
    y = 3*x
    return y
print(f())
```

**- NESTED FUNCTION**

```python
# Nested Function
def square():
    def add():
        x = 3
        y = 4
        z = x + y
        return z
    return add()**2
print(square())
```

**- DEFAULT AND FLEXIBLE ARGUMENTS**

```python
# Default and Flexible Arguments
# Default Argument
def f(a, b = 1, c = 4):     #if we can not write (a) firstly, there is an error!
    x = a + b + c
    return x
print(f(3))
print(f(1,4,7))
```

```python
# Flexible Argument
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,4,7,10)

def f(*kwargs):
    """ print key and value of dictionary"""
    for key, value in kwargs.items():
        print(key, " ", value)
# (!)f(country = 'usa', brand = 'chevrolet')
```

**- LAMBDA FUNCTION**

```python
# Lambda Function
square = lambda x: x**2
print(square(5))
total = lambda x,y,z: x+y+z
print(total(1,3,5))
```

**- ANONYMOUS FUNCTION**

```python
# Anonymous Function
number_list = [1,3,5]
y = map(lambda x:x**2, number_list)
print(list(y))
```

**- ITERATION**

```python
# Iteration
name = "gomez"
it = iter(name)
print(next(it))
print(*it)
```

```python
# zip
list1 = [1,3,5,7]
list2 = [2,4,6,8]
z = zip(list1,list2)
print(z)
z_list = list(z)
print(z_list)
```

```python
un_zip = zip(*z_list)
un_list1,un_list2 = list(un_zip)
print(un_list1)
print(un_list2)
print(type(un_list1))
print(type(un_list2))
```

**- LIST COMPREHENSION**

```python
# List Comprehension
num1 = [1,3,5]
num2 = [i+2 for i in num1]
print(num2)
```

```python
# Conditionals on iterable
num1 = [8,10,12]
num2 = [i**2 if i == 10 else i-3 if i < 10 else i+8 for i in num1]
print(num2)
```
