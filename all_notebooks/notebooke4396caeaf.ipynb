```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import os
import missingno as mion
```

```python
data=pd.read_csv("../input/teslainctsla-historical-prices/TSLA.csv")
```

```python
data['Date']=pd.to_datetime(data['Date'])
```

```python
data['Day']=data['Date'].dt.day_name()
```

```python
data['Day'].value_counts()
```

```python
plt.style.use("fivethirtyeight")
fig,axis=plt.subplots(2,2 ,figsize=(20,10))
sns.lineplot(x='Date',y='Open',data=data , hue="Day",ax=axis[0,0])
sns.lineplot(x='Date',y='High',data=data , hue="Day",ax=axis[0,1])
sns.lineplot(x='Date',y='Low',data=data , hue="Day",ax=axis[1,0])
sns.lineplot(x='Date',y='Close',data=data , hue="Day",ax=axis[1,1])
```

```python
plt.figure(figsize=(12,7))
plt.style.use("ggplot")
sns.lineplot(x='Date',y='Volume',data=data ,hue="Day" )
```
