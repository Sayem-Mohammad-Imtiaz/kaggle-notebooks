```python
# importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# importing the data set
df = pd.read_csv('../input/kmeans-clustering/kmeans.csv')
print(df)
```

<p><b> here i will be showing this algo with the age and annual column your assignment will be to perform the same algo with the age~spending score and annual incone~spending score

```python
x = df.iloc[:,2:4].values
print(x)
```

# Applying the elbow method for selecting the best k cluster value

```python
from sklearn.cluster import KMeans
wcss = []
x1 = range(1,11)
for i in x1:
    clus = KMeans(n_clusters=i,init='k-means++',random_state=0)
    clus.fit(x)
    wcss.append(clus.inertia_)
plt.scatter(x1,wcss,c='red')
plt.plot(x1,wcss,c='blue')
plt.show()
```

```python
# training the kmeans object with k=4
cluster = KMeans(n_clusters=4,init='k-means++',random_state=0)
cluster.fit(x)
```

```python
# predicting the output
y_pred = cluster.predict(x)
print(y_pred)
```

```python
y_col = y_pred.reshape(len(y_pred),1)
output = np.append(arr=x,values=y_col,axis=1)
print(output)
```

```python
plt.scatter(x[y_pred==0,0],x[y_pred==0,1],c='yellow',s=80,label='cluster1')
plt.scatter(x[y_pred==1,0],x[y_pred==1,1],c='cyan',s=80,label='cluster2')
plt.scatter(x[y_pred==2,0],x[y_pred==2,1],c='green',s=80,label='cluster3')
plt.scatter(x[y_pred==3,0],x[y_pred==3,1],c='red',s=80,label='cluster4')
plt.scatter(cluster.cluster_centers_[:,0],cluster.cluster_centers_[:,1],
            c='black',s=100,label='centroid')
plt.legend()
plt.xlabel('AGE')
plt.ylabel('Annual Income')
plt.show()
```
