### SVD and（Principal component analysis） PCA belongs to the entry algorithm in matrix algorithm
*PCA uses information measurement indicators. The larger the variance of the sample, the more information the feature carries

```python
#The dimensionality reduction requires a large amount of calculation and runs slowly.but it works on reduce feature.
#PCA compresses existing features, and the reduced-dimensional features are not any original features, 
#but new features combined in some way. Therefore, the dimensionality reduction algorithm represented by PCA is a kind of feature creation

#pca

"""
calss sklearn.decomposition.PCA(n_components=None,copy=True,whiten=False,scd_solver='auto',tol=0.0,iterated_power='auto',random_state=None)

"""
```

```python
import matplotlib .pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris=load_iris()
y=iris.target
x=iris.data
```

```python
# 2-dimensional array, 4-dimensional matrix
x.shape
```

```python
#pca
pca=PCA(n_components=2)
pca=pca.fit(x)#Fitting the model
x_dr=pca.transform(x)#acquire new matrix
x_dr.shape
# x_dr=PCA(2).fit_transform(x)
```

```python
x_dr[y==0,0]# acquire x1,when y==0 ## x_dr[:,0]
```

```python
plt.figure()
plt.scatter(x_dr[y==0,0],x_dr[y==0,1],c='red',label=iris.target_names[0])
plt.scatter(x_dr[y==1,0],x_dr[y==1,1],c='black',label=iris.target_names[1])
plt.scatter(x_dr[y==2,0],x_dr[y==2,1],c='orange',label=iris.target_names[2])
plt.legend()
plt.title('PCA of IRIS dataset')
plt.show()
```

```python
pca.explained_variance_ #Check the amount of information (explainable variance) of each new feature after dimensionality reduction
```

```python
#Interpretable variance contribution rate
pca.explained_variance_ratio_ #Check the percentage of information occupied by each new feature vector to the metadata information after dimensionality reduction
```

```python
pca.explained_variance_ratio_.sum() # new 2 features contain 4 feature information
```

```python
import numpy as np
np.cumsum(pca.explained_variance_ratio_) # Cumulative result
```

```python
pca_line=PCA().fit(x)
plt.plot([1,2,3,4],np.cumsum(pca_line.explained_variance_ratio_))
plt.xticks([1,2,3,4])#Restrict the axes to be integers
plt.xlabel('number of components after dimension reduction')
plt.ylabel('cumculative explained variance')
plt.show()
```

# inverse_transform

```python
faces=fetch_lfw_people(min_faces_per_person=60)
faces.images.shape

x=faces.data
```

```python
faces.data.shape
```

```python
pca=PCA(150)
x_dr=pca.fit_transform(x)
x_dr.shape
```

```python
x_inverse=pca.inverse_transform(x_dr)
x_inverse.shape
```

```python
# We compare the original image with the restored image after dimensionality reduction
fig,ax=plt.subplots(2,10,figsize=(10,2.5)
                   ,subplot_kw={"xticks":[],"yticks":[]}
                   )
for i in range(10):
    ax[0,i].imshow(faces.images[i,:,:],cmap='binary_r')
    ax[1,i].imshow(x_inverse[i].reshape(62,47),cmap='binary_r')
ax
```

* Because the data after dimensionality reduction is upgraded again, 
  it cannot make up for the information lost in the process of dimensionality reduction, so the image is vague

* After face recognition of ID card information being scanned, a lot of information has been stored.
* However,we can only extract less information  by dimensionality reduction,
* and then a small amount of information will be used for comparison with real people,
* Even so it can also have a high matching accuracy

# Use PCA for noise filtering

```python
from sklearn.datasets import load_digits
digts=load_digits()
digts.data.shape
```

```python
digts.data
```

```python
digts.images.shape
```

```python
set(digts.target.tolist())
```

```python
def plot_digts(data):
    # The structure of data must be (m, n) and n must be divided into (8, 8)
    fig,axes=plt.subplots(4,10,figsize=(10,4)
                   ,subplot_kw={"xticks":[],"yticks":[]}
                   )
    for i,ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8,8),cmap='binary')
```

```python
plot_digts(digts.data)
```

```python
# add noise
np.random.RandomState(42)
noisy=np.random.normal(digts.data,2) #Sampling from the digit data set,and the selected samples meet the normal distribution ,variance=2
plot_digts(noisy)
```

```python
# Noise reduction
pca=PCA(0.5,svd_solver='full').fit(noisy)# The data after dimensionality reduction contains more than 50% of the original data information
x_dr=pca.transform(noisy)
x_dr.shape
```

```python
without_noisy=pca.inverse_transform(x_dr)
without_noisy.shape
```

```python
plot_digts(without_noisy)
```

```python
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

#### import data

```python
data=pd.read_csv('../input/mnist-digit-recognizer/train.csv')
data.shape
```

```python
X=data.iloc[:,1:]
y=data.iloc[:,0]
print(X.shape,y.shape)
```

#### Draw the cumulative variance contribution rate curve, find the best dimension range after dimensionality reduction

```python
pca_line=PCA().fit(X)
plt.figure(figsize=[20,5])
plt.plot(np.cumsum(pca_line.explained_variance_ratio_))
plt.xlabel('number of components after dimension reduction')
plt.ylabel('cumulative explained variance ratio')
plt.show()
# so 50 features contain almost 80% information
```

#### learning curve  after dimensionality reduction

```python
score=[]
for i in range(1,101,10):
    x_dr=PCA(i).fit_transform(X)
    once=cross_val_score(RFC(n_estimators=10,random_state=0)
                        ,x_dr,y,cv=5).mean()
    score.append(once)
plt.figure(figsize=[20,5])
plt.plot(range(1,101,10),score)
plt.show()
# when n_compoment==20 we can accquire over 0.9 score
```

```python
score=[]
for i in range(10,25):
    x_dr=PCA(i).fit_transform(X)
    once=cross_val_score(RFC(n_estimators=10,random_state=0)
                        ,x_dr,y,cv=5).mean()
    score.append(once)
plt.figure(figsize=[20,5])
plt.plot(range(10,25),score)
plt.show()
# when n_compoment==21, we got the max score
```

```python
x_dr=PCA(21).fit_transform(X)
cross_val_score(RFC(n_estimators=10,random_state=0),x_dr,y,cv=5).mean()
```

```python
# import n_estimators to 100 ,to see whether the score improve
cross_val_score(RFC(n_estimators=100,random_state=0),x_dr,y,cv=5).mean()
```

```python
# Test on KNN model
from sklearn.neighbors import KNeighborsClassifier as KNN
cross_val_score(KNN(),x_dr,y,cv=5).mean()
```

```python
# find the best k of KNN
score=[]
for i in range(10):
   # x_dr=PCA(21).fit_transform(X)
    once=cross_val_score(KNN(i+1),x_dr,y,cv=5).mean()
    score.append(once)
plt.figure(figsize=[20,5])
plt.plot(range(10),score)
plt.show()
# we can find when k==3 ,we can get best score
```
