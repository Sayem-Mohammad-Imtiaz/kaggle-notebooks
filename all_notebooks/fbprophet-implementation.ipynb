```python
import pandas as pd
from prophet import Prophet
```

```python
df = pd.read_csv("../input/dogecoin-historical-data-03082019-05052021/Doge_historical_data.csv",usecols=["Date","Open"])
df.head()
```

```python
df.columns = ['ds','y']
df.head()
```

```python
m = Prophet()
m.fit(df)
```

```python
future = m.make_future_dataframe(periods=365)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = m.plot(forecast)
```

```python
fig2 = m.plot_components(forecast)
```
