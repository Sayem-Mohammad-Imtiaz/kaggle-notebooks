```python
#!/usr/bin/env python
# coding: utf-8

# In[1]:


import pandas as pd


# In[2]:


import numpy as np


# In[3]:


import matplotlib.pyplot as plt


# In[10]:


#uploading company list as a list.
path= "../input/stocks-analysis"
company_list=['AAPL_data1.csv','GOOG_data1.csv','MSFT_data1.csv','AMZN_data1.csv']
all_data=pd.DataFrame()
for file in company_list:
    current_df=pd.read_csv(path+'/'+file)
    all_data=pd.concat([all_data,current_df])
              


# In[5]:


company_list


# In[15]:


all_data.shape


# ANALYSE CLOSING PRICE OF ALL THE STOCK

# In[18]:


#storing stock name
tech_list=all_data['Name'].unique()


# In[19]:


tech_list


# In[22]:


#converting date from string to date time
all_data.dtypes


# In[24]:


all_data['date']= pd.to_datetime(all_data['date'])


# In[25]:


all_data.dtypes


# In[27]:


#plotting closing time of each stock
plt.figure(figsize=(20,12))
for i,company in enumerate (tech_list,1):
    plt.subplot(2,2,i)
    df=all_data[all_data['Name']==company]
    plt.plot(df['date'],df['close'])
    plt.xticks(rotation='vertical')
    plt.title(company)


# plot total daily volume traded

# In[31]:


get_ipython().system('pip install plotly')
import plotly.express as px


# In[33]:


for company in tech_list:
    df=all_data[all_data['Name']==company]
    fig=px.line(df,x='date', y='volume', title=company)
    fig.show()


# Analysing daily price change

# In[35]:


df=pd.read_csv("../input/stocks-analysis/AAPL_data1.csv")


# In[37]:


df.head()


# In[38]:


#adding column price change
df['daily_price_change']=df['close']-df['open']


# In[40]:


df.head()


# In[43]:


df['1_day_%_change']=((df['close']-df['open'])/df['close'])*100


# In[44]:


df.head()


# In[46]:


fig=px.line(df,x='date', y='1_day_%_change', title=company)
fig.show()


# Anlyse monthly mean of close feature

# In[48]:


df2=df.copy()


# In[50]:


df2.dtypes


# In[58]:


df2['date']=pd.to_datetime(df2['date'])
df2['date']


# In[59]:


#setting date as index so that we can easily find and all for intervals
df2.set_index('date',inplace=True)


# In[60]:


df2.head()


# In[62]:


df2['2013-02-08':'2013-02-12']


# In[64]:


#plotting mean closing time trend ## resample is used to make the bin size monthly
df2['close'].resample('M').mean().plot()


# In[67]:


df2['close'].resample('Y').mean().plot(kind='bar')


# Are values of Tech companies co-related?

# In[71]:


#upploading data set of tech companies amazon, google, microsoft, apple
amzn=pd.read_csv("../input/stocks-analysis/AMZN_data1.csv")
amzn.head()


# In[72]:


goog=pd.read_csv("../input/stocks-analysis/GOOG_data1.csv")
goog.head()


# In[73]:


msft=pd.read_csv("../input/stocks-analysis/MSFT_data1.csv")
msft.head()


# In[74]:


aapl=pd.read_csv("../input/stocks-analysis/AAPL_data1.csv")
aapl.head()


# In[79]:


close=pd.DataFrame()


# In[80]:


close['aapl']=aapl['close']
close['amzn']=amzn['close']
close['msft']=msft['close']
close['goog']=goog['close']


# In[81]:


close.head()


# In[88]:


#to find co-relation :
import seaborn as sns
sns.pairplot(data=close)


# In[89]:


#so in the above graph we can see that amazon and microsoft prices are very much corelated. (thinner and straighter the line more the corelation)


# In[91]:


sns.heatmap(close.corr(),annot=True)


# In[92]:


#in the above graph same correlation in the form of heatmap.


# Analyse daily return of each stock

# In[95]:


aapl.head()


# In[96]:


data=pd.DataFrame()


# In[99]:


data['aapl_change']=((aapl['close']-aapl['open'])/aapl['close'])*100
data['goog_change']=((goog['close']-goog['open'])/goog['close'])*100
data['msft_change']=((msft['close']-msft['open'])/msft['close'])*100
data['amzn_change']=((amzn['close']-amzn['open'])/amzn['close'])*100


# In[100]:


data.head()


# In[101]:


sns.pairplot(data=data)


# In[102]:


#histogram area denotes no relation as compairing same companies doesnt make sense


# In[103]:


sns.heatmap(data.corr(),annot=True)


# In[104]:


#same correlation


# Value at risk analysis for tech companies

# In[108]:


#distribution plot
sns.distplot(data['aapl_change'])


# In[110]:


#checking standard deviation
##68%of entire data
data['aapl_change'].std()


# In[111]:


#2nd std deviation
###95%of entire data
data['aapl_change'].std()*2


# In[112]:


#3rd std deviation
####99.17% of entire data
data['aapl_change'].std()*3


# In[113]:


#so we can say 99.17% of time apple returns will be between -3.561 to +3.561


# In[117]:


#using quantile statement
#90% of the time of the worst loss will not be more than this
data['aapl_change'].quantile(0.1)


# In[119]:


data.describe().T


#   end
#     

# In[ ]:





# In[ ]:
```
