Initial data preprocessing and Data Preparation

```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import datetime 
import calendar
import plotly.express as px

```

```python
 df = pd.read_csv(r'../input/covid-world-vaccination-progress/country_vaccinations.csv')
df.head()
```

```python
df.date = pd.to_datetime(df.date)
df['month']= df.date.dt.month
df['day']= df.date.dt.day
```

```python
df['Week_day']=df.date.dt.weekday
```

```python
df.head()
```

```python
ind = df.loc[df['country'] == 'India']
```

DATA ANALYSIS

India second most populous country in the world and world's largest democracy has a huge task of vaccinating its population of over 1.3 billion people.
The country officially accepted two vaccines namely Oxford-Astrazaneca and Covaxin on 1st of January 2021.
Though vaccination started from 15th January with first phase aimed for its healthcare workers who amount to more than 10 million. The frontline workers will soon follow who are estimated around 20 million. 

```python

plt.figure(figsize=(15,8))
sns.lineplot(x="date", y="total_vaccinations",data=ind)
```

With initial hesitation and rumours the vaccine benefitiaries soon touched 10 million in a span of more than 30 days.

```python
plt.figure(figsize=(15,8))
sns.barplot(x="Week_day", y="daily_vaccinations_raw",data=ind)
```

The beneficiaries increased substantially in February compared to January naturally due to increased awareness, positive word of mouth and fear among the general public for a possible second wave like the western nations.

Also unorganised businesses form the backbone of the country's economy who severely were affected due to harsh lockdowns and social distancing curbs.

```python
sns.barplot(x="month", y="total_vaccinations",data=ind)
```

```python
map_plot_df = pd.concat([pd.Series(row['iso_code'], row['vaccines'].split(','))              
                         for _, row in df.iterrows()]).reset_index()
map_plot_df.columns = ['vaccine', 'iso_code']

#strip white spaces
map_plot_df['vaccine'] = map_plot_df['vaccine'].str.strip(' ')

map_plot_df.vaccine.value_counts()

```

The map is interactive play around to get your information

India houses some of the largest vaccine manufacturers in the world.With a price advantage Indian vaccines are considered as game changers for the middle and lower middle income countries who also face infrastructural issues pertaining to cold storage.

Some of the major global vaccines have collaborated with Indian manufacturers:
Oxford-Astrazeneca: Serum Institute of India
Sputnik : Dr Reddy Labrorarities Limited

So India remains a key link in supply chain of vaccines globally.


```python
fig = px.choropleth(locations = map_plot_df.iso_code, 
                    color = map_plot_df.vaccine,
                    color_continuous_scale = 'rainbow',
                    title = 'Where is each vaccine used?', 
                   )
fig.show()
```

India a country with second largest case count stands behind United Kingdom in total vaccine benefitiaries. 
The country claims to be fastest vaccinator globally. Though it stands far behind its population peers China and Uinted States who are first and third most populous country on the planet. 
Possible reasons being:
1. Poor Health Infrastructure
2. Less Awareness and spread of rumours
3. Initial hiccups due to failure of registration platform.



```python
df.groupby(['country']).agg({'total_vaccinations':'max'}).sort_values(by='total_vaccinations', ascending=False)[:10].plot.barh(figsize=(15,5))
plt.title('TOTAL VACCINATIONS PER COUNTRY - Top 10')
```

```python

df.groupby(['country']).agg({'daily_vaccinations':'mean'}).sort_values(by='daily_vaccinations', ascending=False)[:10].plot.barh(figsize=(10,5))
plt.title('DAILY VACCINATIONS TOP 10')
```

```python
df.groupby(['country']).agg({'daily_vaccinations_per_million':'mean'}).sort_values(by='daily_vaccinations_per_million', ascending=False)[:10].plot.barh(figsize=(15,5))
plt.title('DAILY VACCINATIONS PER MILLION - Top 10')
```

```python
ind.groupby(['country']).agg({'daily_vaccinations_per_million':'mean'}).plot.barh(figsize=(10,5))
plt.title('DAILY VACCINATIONS PER MILLION INDIA')
```

The Pace of vaccination seems slow as compared to other world countries. All though domestic production can help India source vaccines without delays.
An increase in awareness possibly through political and cultural influencers can significantly help the country. 

India aims to start vaccinating people above the age of 50 from 1st March 2021.They amount to 300 million. An urgent increase in awareness and publicity can help increase benefitiaries as well avoid wastage of vaccinations due to under utilization.
Which can be provided to needy countries and other UN World health missions.
