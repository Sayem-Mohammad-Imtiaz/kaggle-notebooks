### Missing values

```python
dict_isna = {"Variable": [], "Num_NaN": [], "Percent_NaN": []}
for column, na_num in zip(df_app.columns, df_app.isna().sum().to_list()):
    dict_isna["Variable"].append(column)
    dict_isna["Num_NaN"].append(na_num)
    dict_isna["Percent_NaN"].append(np.around((na_num/df_app.shape[0])*100, decimals=2))
df_app_isna = pd.DataFrame(dict_isna)
df_app_isna.sort_values(by="Num_NaN", ascending=False)
```

### Outliers
Remove the most extreme outliers while keeping rows containing NaN

```python
cutoff = 0.99
drop_idx = []
for column, dtype, in zip(df_app.columns, df_app.dtypes):
    if dtype == "float64":
        drop_idx += list(df_app[df_app[column]>df_app[column].quantile(cutoff)].index)
    if dtype == "int64" and df_app[column].unique().shape[0]>10:
        drop_idx += list(df_app[df_app[column]>df_app[column].quantile(cutoff)].index)
drop_idx = list(dict.fromkeys(drop_idx)) # Remove duplicates
df_app.drop(drop_idx, inplace=True)
```

### Correlations
With a total of 122 features visualizations of the correlations are difficult to plot.<br>
I decided to use tables with pairwise correlations instead.<br>
A first simplification is that i'll only analyze numerical features (and drop features with dtype "object").
Some features are given with their mean, median and mode. As we can expect to see similar correlations for all these variations i decided to use only one of these variable types per feature and drop the rest.

```python
numeric_columns = df_description[(df_description.Variable_dtype=="int64") | (df_description.Variable_dtype=="float64")].index.to_list()
```

```python
df_app_corr = df_app[numeric_columns].corr(method="spearman")
```

```python
dict_correlations = {"Variable1": [], "Variable2": [], "Correlation": []}
columns = list(df_app_corr.columns)
for i, column in enumerate(columns):
    if i==len(columns):
        break
    for j, column2 in enumerate(columns[i+1:]):
        if ("_".join(column.split("_")[:-1])=="_".join(column2.split("_")[:-1])) and ((column.split("_")[-1]=="MEDI") or (column.split("_")[-1]=="AVG") or (column.split("_")[-1]=="MODE")) and ((column2.split("_")[-1]=="MEDI") or (column2.split("_")[-1]=="AVG") or (column2.split("_")[-1]=="MODE")):
            pass
        else:         
            dict_correlations["Variable1"].append(column)
            dict_correlations["Variable2"].append(column2)
            if np.isnan(df_app_corr[column][column2]):
                dict_correlations["Correlation"].append(0.)
            else:
                dict_correlations["Correlation"].append(df_app_corr[column][column2])
```

```python
df_app_corr_disp = pd.DataFrame(dict_correlations)
```
