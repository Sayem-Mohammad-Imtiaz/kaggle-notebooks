## What are the most common types of deep sea coral?

```python
coral_by_frequency = (
    deep_sea_coral
    .groupby("ScientificName")
    .size()
    .sort_values(ascending=False)
)

coral_by_frequency.head(10)
```

Here are some short descriptions with links for more information. Oddly enough, some species of coral are hard to find on Wikipedia. 

1. [Porifera][1] - The sponge, a multicellular organism with bodies full of pores and channels to circulate water through them.
2. Heteropolypus ritteri - A deep-sea mushroom coral originally known by [_Anthomastus ritteri_][2], which can be found in the northern Mid-Atlantic waters.
3. [Lophelia pertusa][3] - A slow growing, cold water coral that is a habitat of choice of many deep sea fish. A fairly high-profile deep-sea coral due to unsustainable fishing and oil practices in the Northern Atlantic. Protected under [CITES article II][4]. 
4. [Pennatulacea][5] - Sea pens, named after a distinctive resemblance to writing quills. It is a soft coral with 14 families, 35 genera, and over 200 described species with a cosmopolitan distribution. 
5. [Stylaster sp.][6] - Also known as "lace coral", it can be found in deep waters around caves and downward ledges. It is a kind of colonial [hydrozoa][7], individually small predatory animals.
6. [Hexactinellida][8] - The "glass sponge", a sponge with a skeleton with four/six-pointed spicules common to Antartic and Northern Pacific waters. They have the unique ability to react quickly to stimuli by conducting electrical pulses across their bodies. They form reefs.
7. [Gorgonacea][9] - The accepted name is actually [Alcyonacea][10]. Known as "sea fans" or "sea whips", closely related to the sea pen. They do not produce calcium carbonate. Their size, shape, and appearance are correlated to their location, most likely to maximize the efficiency of their filter feeding process.
8. [Paragorgia sp.][11] - Named "bubblegum coral" due to their large branching fan structure with bulbous tips. They are found in the northern Atlantic Ocean living at depths of 200-1400 meters between 2-8C, and first described by Carl Linnaeus. Prefers locations with strong currents. 
9. [Farrea occa][12] - Known as the "reversed glass sponge", it has been described as a globular, shrubby structure. An [interesting description][13] is provided by Dr. J. Anderson from a specimen obtained from the Central Kuroshio Current in 1885.
10. [Funiculina sp.][14] - A genus of sea pens known as "ムチウミサボテン" or "whip sea cactus". 

Some class of coral has been estimated to around since Cambrian Period, an era known for the [explosion of life 541 million years ago](https://en.wikipedia.org/wiki/Cambrian_explosion) where most major animal phyla appeared. The general method of subsistence seems to be a form of filter feeding, which requires the mechanical motion of ocean currents to bring food.
