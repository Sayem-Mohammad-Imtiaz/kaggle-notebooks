```python
df = pd.read_csv('/kaggle/input/blood-type-distribution-by-country/bloodtypes.csv',index_col=[0]);
```

```python
df.head()
```

```python
df.info()
```

```python
#Change the datatype for population
df['Population']=df['Population'].str.replace(',', '').astype(float)
```

```python
col = df.columns[2::]
col.astype(str)
col
```

```python
#change the datatype for the rest of the columns and convert % to number
for i in range(len(col)):
    df[col[i]] = df[col[i]].str.replace("%","").astype(float)
    df[col[i]] = df[col[i]].apply(lambda x: (x/100.0)*df['Population']);
```

```python
#checking the Datatypes
df.info()
```
