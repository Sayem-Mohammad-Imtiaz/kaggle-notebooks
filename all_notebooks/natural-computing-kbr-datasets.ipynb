```python
def load_dataset(path):
    #Load dataset
    dataframe = pd.read_csv(path)
    display(dataframe)
    
    #Features in dataset
    #print(list(dataframe))
    print(dataframe.shape)

    #Datatypes in dataset
    print(dataframe.dtypes)
    
    return dataframe
```

```python
def find_null_vals(dataframe):
    #Number of null values in dataframe
    nr_nulls = dataframe.isnull().sum().sum()
    print(f"There are {nr_nulls} null values")
    if nr_nulls != 0:
#         #Dataframe boolean to see where null values are
#         print(dataframe.isnull())
        #Nullvalues per column
        print(dataframe.isnull().any(axis=0))
        dataframe = dataframe.dropna(how='any',axis=0) 
        print(f'New shape: {dataframe.shape}')
        print(f'There are now: {dataframe.isnull().sum().sum()} null values in the dataframe.')
        
    return dataframe
```

```python
def make_hist(dataframe, rows, columns, tight_layout = True, figsize = (18, 18)):
    # from matplotlib.pyplot import hist as hist
    column = 0
    row = 0
    n_bins = 10
    sns.set()
    #Histograms with distributions for all features
    fig, axs = plt.subplots(rows, columns, sharey=True, tight_layout=tight_layout, figsize = figsize)

    for i in dataframe: 
        axs[row][column].hist(dataframe[i], bins=n_bins)
        axs[row][column].title.set_text(i)
        column += 1
        if column % columns == 0:
            row += 1
            column = 0
```

```python
def Make_One_Hot(dataframe):
    print(dataframe.dtypes)
    no_string_count = 0
    
    for i in dataframe:
        if dataframe[i].dtype == 'O':
            print(f"There are {no_string_count} columns here which do not contain strings (including the previous no string print).") #print this for the columns before this column with a string
            print(f"string found in column {i}")
            ohe = pd.get_dummies(dataframe[i])
            dataframe = pd.concat([dataframe,ohe], axis= 1)
    #pd.merge(left=dataframe, right=ohe, how = 'left', left_index=True,right_index=True)
            dataframe = dataframe.drop(i,axis = 1)
            no_string_count = 0
        else:
            no_string_count += 1
            if no_string_count < 2 :
                print("No string")
    
    print(f"new dataframe shape: {dataframe.shape}")
    return dataframe
```

```python
def split(data, train_size = 0.7, test_size = 0.5):
    test_set = 1-train_size
    x_train, x_test = train_test_split(data, test_size = test_set, random_state=666)
    x_test, x_val = train_test_split(x_test, test_size = test_size, random_state=666)
    
    return x_train, x_test, x_val
```

```python
def preprocess_result(columnval):
    home, away = columnval.split("-")
    if home > away:
        return 1 #home wins
    if away > home: 
        return 2 #away wins
    return 0 #draw
```
