```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from fbprophet import Prophet
```

```python
df = pd.read_excel("/kaggle/input/hosp-beds-data/DC-COVID-19-Data-for-June-8-2021.xlsx",engine = "openpyxl")
```

```python
df_beds = df.loc[10].transpose().reset_index()
```

```python
df_beds.head()
```

```python
df_beds.columns = ['Date','Occupied']
```

```python
df_beds = df_beds.drop(range(0,27),axis = "rows").reset_index(drop = True)
```

```python
df_beds.isna().sum()
```

```python
df_beds.dropna(inplace = True)
```

```python
df_beds['Date'] = pd.to_datetime(df_beds['Date'])
df_beds['Occupied'] = df_beds['Occupied'].astype('int64')
```

```python
plt.figure(figsize = (16,10))
sns.scatterplot(data = df_beds, x = 'Date',y = 'Occupied')
```

```python
df_beds.columns = ['ds','y']
```

```python
model = Prophet()

model.fit(df_beds)
```

```python
df_beds['ds'].tail()
```

```python
future= model.make_future_dataframe(periods=30)
```

```python
future.tail(10)
```

```python
forecast = model.predict(future)
```

```python
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
plt.figure(figsize = (15,6))
fig = model.plot(forecast)
```
