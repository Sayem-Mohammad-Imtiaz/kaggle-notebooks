```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

# Reading Data

```python
crime = pd.read_csv('../input/Crime1.csv')
crime.head()
```

Making Count of Category and plotting

```python
counts = crime.Category.value_counts()
df = pd.DataFrame(counts)
df['name'] = df.index
df.columns = ['Value','name']
df.head()
```

```python
sns.factorplot(x='Value',y='name',data=df,kind='bar',size=4.25, aspect=1.9)
```

# Top 10 most committed crimes

```python
counts = crime.Descript.value_counts()[:10]
df = pd.DataFrame(counts)
df['name'] = df.index
df.columns = ['Value','name']
df.head()
```

```python
sns.factorplot(x='Value',y='name',data=df,kind='bar',size=4.25, aspect=1.9)
```

# Day on which there is most crime?

```python
sns.countplot(crime['DayOfWeek'])
```

# District with most crimes

```python
counts = crime.PdDistrict.value_counts()
df = pd.DataFrame(counts)
df['name'] = df.index
df.columns = ['Value','name']
df.head()
```

```python
sns.factorplot(x='name',y='Value',data=df,kind='bar',size=4.25, aspect=1.9)
plt.xticks(rotation=30)
```

# Checking Resolution

To check if crimes are resolved or not

```python
counts = crime.Resolution.value_counts()
df = pd.DataFrame(counts)
df['name'] = df.index
df.columns = ['Value','name']
df.head()
```

```python
sns.factorplot(x='name',y='Value',data=df,kind='bar',size=4.25, aspect=1.9)
plt.xticks(rotation=30)
```

Looks like most crimes were not resolved.

# Processing date column to get crime hours

```python
from datetime import datetime
def processDate():
    for index,row in crime.iterrows():
        date = datetime.strptime(row['Dates'], '%m/%d/%Y %H:%M')
        crime.loc[index,'hour'] = date.hour
        crime.loc[index,'day'] = date.day
```

```python
processDate()
```

```python
crime.head()
```

# Plotting crime by hour

```python
plt.figure(figsize=(10,6))
sns.distplot(crime['hour'].values,kde=False,rug=False,bins=24,color='green')
```

Looks like crimes have higher rate at mid-night, as expected! (LOL)

# Plotting crime by day of month

```python
plt.figure(figsize=(10,6))
sns.distplot(crime['day'].values,kde=False,rug=False,bins=24,color='blue')
```
