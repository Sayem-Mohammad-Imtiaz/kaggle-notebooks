## Mengecek outlier dengan IQR

```python
# Membuat sebuah fungsi untuk menemukan outlier
from numpy import quantile
def outlier(df_in, col_name):
    q1 = df_in[col_name].quantile(0.25)
    q3 = df_in[col_name].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df_in.loc[(df_in[col_name] <= fence_low) | (df_in[col_name] >= fence_high)]
    return df_out
```

```python
# Menerapkan fungsi memisahkan outlier dan mendeklarasikannya kedalam suatu variabel
high_kematian = outlier(df2, 'Deaths / 100 Cases').reset_index() # melihat outlier pada kolom 'Kematian'
high_kematian.head()
```

Ternyata tidak ada outliers pada kolom 'Deaths / 100 Cases'
