<h2>Important Note</h2>

<h6>In our Price column all values are numerical except Free one so need to convert it into the 0 so that it can make our visualization easier :)</h6>

Step -1 Removing the ₹ sign

```python
data['Price']=data['Price'].str.replace(',', '')
data['Price']=data['Price'].str.replace('₹', '')
```

Step - 2 ( Converting the Free into 0 and converting the values into integer)

```python
def con(x):
    if x=='Free':
        return 0
    else:
        return x
```

```python
data['Price']=data['Price'].apply(lambda x:con(x))
```

```python
data['Price']=pd.to_numeric(data['Price'],errors='coerce')
```
