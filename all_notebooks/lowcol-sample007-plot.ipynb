```python
def hexbin(x, y, color, **kwargs):
    cmap = sns.light_palette(color, as_cmap=True)
    plt.hexbin(x, y, cmap=cmap, **kwargs)
def scatter(x, y, color, **kwargs):
    plt.scatter(x, y, marker='.')
```

```python
img_cnt = 0
def save_img():
    global img_cnt
    img_cnt += 1
    img_name = 'img_{0:03}.png'.format(img_cnt)
    print(img_name)
    plt.savefig(img_name)
```

```python
df1 = pd.DataFrame(df_wgt_col.copy().values)
df2 = pd.DataFrame(df_wgt_user.copy().values)
df = pd.concat([df2, df1])

plt.figure(figsize=(12,12))
ax = sns.kdeplot(data=df2, x=0, y=1,
                 levels=5, thresh=.1)
ax.scatter(df1.values[:,0], df1.values[:,1], c='darkgreen', s=200, marker='X')
save_img()
```

```python
df1 = pd.DataFrame(df_wgt_col.copy().values)
df2 = pd.DataFrame(df_wgt_user.copy().values, columns=['cx', 'cy'])
df = pd.concat([df2, df1])

g = sns.lmplot(data=df2, x='cx', y='cy', fit_reg=False, height=10, scatter_kws={'s':100})
ax = g.ax
ax.scatter(df1.values[:,0], df1.values[:,1], c='darkgreen', s=200, marker='X')
for ii, ee in enumerate(df1.iterrows()):
    ax.text(ee[1][0], ee[1][1], s=df_wgt_col.index.values[ii], size=20)
save_img()
```
