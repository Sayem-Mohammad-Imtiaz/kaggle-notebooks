```python
from sklearn import tree
from sklearn import preprocessing
import pydotplus # To create our Decision Tree Graph
from IPython.display import Image
```

lalu kita panggil dataset yang akan digunakan dalam sesi ini...

```python
data_train = pd.read_csv('/kaggle/input/main-tenis/play_tennis_train.csv')
df = pd.DataFrame(data_train,columns=['outlook','temp','humidity','wind','play'])
x_train = pd.get_dummies(df[['outlook','temp','humidity','wind']])
y_train = df[['play']]
```

```python
clf = tree.DecisionTreeClassifier()
nbtrain = clf.fit(x_train, y_train)
```

```python
print(tree.export_graphviz(nbtrain, None))

#Create Dot Data
dot_data = tree.export_graphviz(nbtrain, out_file=None, feature_names=list(x_train.columns.values), 
                                class_names=['No', 'Yes'], rounded=True, filled=True) #Gini decides which attribute/feature should be placed at the root node, which features will act as internal nodes or leaf nodes#Create Graph from DOT data
graph = pydotplus.graph_from_dot_data(dot_data)

# Show graph
Image(graph.create_png())
```
