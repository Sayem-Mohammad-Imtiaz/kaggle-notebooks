## I am a firm believer of simplicity.
## Rather than trying highly complex model a simple linear model can give a very good accuracy and it helps beginners too.
### So here comes our beloved Logistic Regression by which we can classify whether a text is spam or ham.

```python
import pandas as pd
import numpy as np
df = pd.read_csv("../input/spam-email/spam.csv")
df.head()
```

```python
df.Category = df.Category.apply(lambda x: 1 if x == 'spam' else 0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.tokenize import word_tokenize
from sklearn import metrics
from sklearn import model_selection
from sklearn import linear_model
```

```python
if __name__ == "__main__":
    #create a new column called kfold and fill it with -1
    df['kfold'] = -1
    #randomize the rows of the data
    df = df.sample(frac = 1).reset_index(drop = True)
    
    y = df.Category.values
    #initialize the kfold class 
    kf = model_selection.StratifiedKFold(n_splits = 5)
    #fill the new kfold column
    for feature, (train, validation) in enumerate(kf.split(X =df, y =y)):
        df.loc[validation, 'kfold'] = feature
    #let's go over the folds created
    for fold_ in range(5):
        #temporary dataframes for train and test
        train_df = df[df.kfold != fold_].reset_index(drop = True)
        test_df = df[df.kfold == fold_].reset_index(drop = True)
        
        #initialize count vectorizer
        count_vec = CountVectorizer(tokenizer = word_tokenize, token_pattern = None)
        
        #fit the count on trainig data Message
        count_vec.fit(train_df.Message)
        #transform
        X_train = count_vec.transform(train_df.Message)
        X_test = count_vec.transform(test_df.Message)
        
        #initialize Logistic regression
        model = linear_model.LogisticRegression()
        #fit the model
        model.fit(X_train, train_df.Category)
        preds = model.predict(X_test)
        
        #calculate accuracy
        accuracy = metrics.accuracy_score(test_df.Category, preds)
        
        print(f"Fold: {fold_}")
        print(f"Accuracy: {accuracy}")
        print("")
```

```python
print(f"The mean accuracy is: {accuracy.mean()}")
```
