# Data cleaning

**The ````"Age Group"```` column has a value `"25 and older"`. This value also includes the values present for "25 to 39", "50 to 64" and "65 and older". Hence I have removed the rows that contain "25 and older" as a value for "Age Group**

**Also, in the "Sex" column, "Total" value includes both Male and Female. Hence it is of not much use to us.**

`````python
df = df[df["Age Group"] != "25 and older"]
`````

`````python
df = df[df["Sex"] != "Total"]
`````

`````python
df.head()
`````

**Right now, the value of the columns `"Bachelor's Degree Holders"`, `"Science and Engineering"`, `"Science and Engineering Related "`, `"Business"`,`"Education"` ,"	Arts, Humanities and Others" are `strings`** 

Hence we need to convert the strings into integers for better analysis.

`````python
# function to convert the data into integer types
def process_data(data):
    data = data.split(",")
#     print(data)
    data = "".join(data)
    return int(data)
`````

`````python
columns = df.columns
`````

`````python
columns = columns[3:]
`````

`````python
columns
`````

`````python
for col in columns:
    df[col] = df[col].apply(process_data)
`````

`````python
df.head()
`````

`````python
sns.countplot(x="Age Group", data=df)
`````

We have equal number of candidates in all the three age groups

`````python
sns.countplot(y="Sex", hue="Age Group", data=df)
`````

We have equal number of male and female candidates in the three age groups

We will look at the data, State wise, grouped by age and sex in that order.

`````python
df.groupby(['State']).sum().style.background_gradient(cmap='Blues').set_table_styles([{'selector' : '',
                            'props' : [('border',
                                        '2px solid darkblue')]}])
`````

**From the above table, we can see that, most of the degree holders are concentrated in the states of California, Florida, New York, Texas**

**Next, we will look at the data state wise , grouped by age groups and sex, in that order.**

`````python
df.groupby(['State','Age Group', "Sex"]).sum().style.background_gradient(cmap='Blues').set_table_styles([{'selector' : '',
                            'props' : [('border',
                                        '2px solid darkblue')]}])
`````

`````python
i=1
plt.figure(figsize=(15,10))
for col in columns:
    
    plt.subplot(2,3,i)
    sns.barplot(x="Age Group", y=col, hue="Sex", data=df, ci=False);
    plt.title(col)
    i+=1
plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)
plt.show()
`````

* We can see that more females are bachelor's degree holder than male students.
* More Male students majored in Science and Engineering and Business than female students.
* More Female students majored in science and engineering related fields, Education and Arts, Humanities and other fields than male students.
