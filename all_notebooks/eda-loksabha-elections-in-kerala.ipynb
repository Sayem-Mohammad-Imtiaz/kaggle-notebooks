```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('../input/india-loksabha-elections-data19622019/Loksabha_election_data_1962-2019')
df.head()
```

```python
df['margin']=df['margin'].str.replace(',', '').astype(float)
```

```python
# Using group by to group elements based on groups
yr=df.groupby('year')
yr
```

```python
#Printing first elements of groups formed
year=yr.first()
years=year.index
year
```

```python
# Showing the results in the year 2019
yr.get_group(2019)
```

```python
# using group by to group based on 3 columns ; state,year and paty
styr = df.groupby(['state', 'year'])
styr.first()
```

```python
styr.count()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# creating a coustom colour pallete
# visit http://colormind.io/ to get color codes you neeed
top='#9E1C00'
mid='#F8E506'
low='#1CA4A4'
```

```python
# list of colors to use
col=[low,mid,top]
```

```python
#plotting bar chart
k19.plot(x='Pc_name',y=['margin','electors','votes'],kind='bar',figsize=(16,8.5),width=.9,color=col)
plt.show
```

```python
co19=k19.groupby('party').count()
```

```python
exp=[]
gg=co19.index
for i in gg:
    print(i)
    exp.append(.03)
exp
```

```python
tmc='#FFF900'
bsp='#31a5fb'
bjp='#ff8c00'
cpi='#ff009f'
cpm='#ff0000'
inc='#083df3'
ncp='#4e2c81'

lstcol=[tmc,bsp,bjp,cpi,cpm,inc,ncp]
```

```python
explode = (0.3, 0.0, 0.3, 0.3, 0.3)

# Wedge properties
wp = { 'linewidth' : 2, 'edgecolor' : "white" }

  
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:d} ".format( absolute)

# Creating plot
fig, ax = plt.subplots(figsize =(20, 7))
wedges, texts, autotexts = ax.pie(co19.votes, 
                                  autopct = lambda pct: func(pct, co19.votes),
                                  explode = explode, 
                                  labels = co19.index,
                                  shadow = True,
                                  colors = lstcol,
                                  startangle = 180,
                                  wedgeprops = wp,
                                  textprops = dict(color ="#173A40"))

# Adding legend
ax.legend(wedges, co19.index,
          title ="Parties",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 

plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Political parties and their seats in 2019")
  
# show plot
plt.show()
```

```python
#combining the above visualisations to show the results in kerala over the years
print('Loksabha elections in kerala - an overview')
for yr in years:
    #display('kerala_'+str(yr))
    n1=pd.DataFrame()
    n1=styr.get_group(('Kerala',yr))
    #display(n1)
    
    #plotting bar chart
    n1.plot(x='Pc_name',y=['margin','electors','votes'],kind='bar',figsize=(16,8.5),width=.9,color=col)
    plt.show
    print(yr)
    
    co19=n1.groupby('party').count()
    
    explode=[]
    exp=co19.index
    for i in exp:
        explode.append(0.1)
    explode.pop()
    explode.append(0)
        
    
    #explode = (0.3, 0.0, 0.3, 0.3, 0.3)

    # Wedge properties
    wp = { 'linewidth' : 2, 'edgecolor' : "white" }

  
    # Creating autocpt arguments
    def func(pct, allvalues):
        absolute = int(pct / 100.*np.sum(allvalues))
        return "{:d} ".format( absolute)

    # Creating plot
    fig, ax = plt.subplots(figsize =(20, 7))
    wedges, texts, autotexts = ax.pie(co19.votes, 
                                  autopct = lambda pct: func(pct, co19.votes),
                                  explode = explode, 
                                  labels = co19.index,
                                  shadow = True,
                                  colors = lstcol,
                                  startangle = 180,
                                  wedgeprops = wp,
                                  textprops = dict(color ="#173A40"))

    # Adding legend
    ax.legend(wedges, exp,
          title ="Parties",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
 

    plt.setp(autotexts, size = 8, weight ="bold")
    ax.set_title("Political parties and their seats in "+str(yr))
  
    # show plot
    plt.show()
    
    
    
```

```python
seatyr=df.groupby(['party','year']).count()
```

```python
seatyr.index
```

```python
seatyr['votes']
```

```python
md = seatyr.reset_index()
md.head(3)
```

```python
seats=md.drop(['Pc_name','type','candidate_name','electors','turnout','margin','margin%','state','votes'],axis=1)
seats.head()
```

```python
seats.rename(columns={'no':'seats'},inplace=True)
seats
```

```python
national_parties=['Indian National Congress','Bharatiya Janta Party','All India Trinamool Congress','Bahujan Samaj Party','Communist Party Of India','Communist Party Of India (MARXIST)','Nationalist Congress Party']
```

```python
bjp_con=seats.loc[seats['party'].isin(national_parties)]
```

```python
import seaborn as sns
```

```python
tmc='#FFF900'
bsp='#31a5fb'
bjp='#ff8c00'
cpi='#ff009f'
cpm='#ff0000'
inc='#083df3'
ncp='#4e2c81'

lstcol=[tmc,bsp,bjp,cpi,cpm,inc,ncp]
```

```python
fig, ax = plt.subplots(figsize=(12, 7))
sns.lineplot(data=bjp_con, x="year", y="seats", hue="party",palette=lstcol,linewidth = 2.5)
plt.legend( bbox_to_anchor=(1.02, 1),loc='upper left', borderaxespad=0, title='Party')
plt.title('Trend of Loksabha seats over the years of Recognized National Parties')
plt.show
```
