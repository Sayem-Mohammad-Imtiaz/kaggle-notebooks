# I mapped fast food restaurants just in case I visit Anaheim and Pittsburgh to watch baseball games of Ohtani of Angels and Tsutsugo of Pirates ! 

```python
df=pd.read_csv('/kaggle/input/fast-food-restaurants-across-us/Fast_Food_Restaurants_US.csv')
```

```python
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import folium
```

```python
df_ana=df[df['city']=='Anaheim']
df_pitts=df[df['city']=='Pittsburgh']
```

```python
df_ana.head()
```

```python
df_pitts.tail()
```

```python
map = folium.Map(location=[33.788917,-117.913451], zoom_start=12)
 
for i, r in df_ana.iterrows():
    folium.Marker(location=[r['latitude'], r['longitude']], popup=r['name']).add_to(map)
    
map.save("map_ana.html")
```

```python
map = folium.Map(location=[40.499500,-79.852900	], zoom_start=12)
 
for i, r in df_pitts.iterrows():
    folium.Marker(location=[r['latitude'], r['longitude']], popup=r['name']).add_to(map)
    
map.save("map_ana.html")
```
