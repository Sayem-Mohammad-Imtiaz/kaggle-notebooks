Suicide rates and graph according to gender

```python
suicide_rate_df.groupby("sex")["suicides_no"].mean()
```

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-whitegrid") 
import seaborn as sns
plt.figure(figsize=(8,5))
sns.barplot(x = "sex", 
            y = "suicides_no", 
            data = suicide_rate_df)
plt.xlabel('Gender')
plt.ylabel('Number of Suicide')
plt.title('Numbers of Suicide by Gender')
plt.show()
```

Countries with highest no of suicide

```python
 suicide_rate_df.groupby("country", as_index = False)["suicides_no"].mean().sort_values(by="suicides_no",ascending = False).head()
```

Average number of sucide per year

```python
yearMeanSN =  suicide_rate_df.groupby("year", as_index = False)["suicides_no"].mean()
plt.figure(figsize=(15,6))
sns.pointplot(x = "year", 
              y = "suicides_no", 
              data = yearMeanSN)
plt.xlabel('Year')
plt.ylabel('Average Number of Suicide')
plt.title('Average Suicide Numbers by Years')
plt.show()
```

Suicides number by age of person

```python
ageData = suicide_rate_df.groupby("age", as_index = False)["suicides_no"].mean().sort_values(by="suicides_no",ascending = False)
plt.figure(figsize=(8,5))
sns.barplot(x = "age", 
            y = "suicides_no", 
            data = ageData)
plt.xlabel('Age Group')
plt.ylabel('Number of Suicide')
plt.title('Suicide Numbers by Age Groups')
plt.show()
```
