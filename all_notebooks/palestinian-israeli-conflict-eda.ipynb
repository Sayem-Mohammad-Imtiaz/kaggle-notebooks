<h1 style="color:black;background-color:white;font-family:Montserrat;font-size:175%;text-align:left;font-weight:Bold">Data preprocessing ( Filling Nulls and fix the data ) for Data sets</h1>

### Body Count Data 

```python
#=======================
df_body=df_body.fillna(0)
#===================
df_body['Month']=df_body['Month'].apply(lambda x:'MAY' if x == 0 else x )
df_body['Year']=df_body['Year'].apply(lambda x:2021 if x == 0 else x )
df_body['Year']=df_body['Year'].apply(lambda x:2021 if x == 0 else x )
#==========================
df_body['Year']=df_body['Year'].apply(lambda x:int(x))
#==================================
df_body['Month']=df_body['Month'].apply(lambda x:x.capitalize())
#================================
df_body["Palestinians Injuries"].iloc[165]=0
df_body["Israelis Injuries"].iloc[165]=0
df_body["Palestinians Injuries"].iloc[203]=0
df_body["Israelis Injuries"].iloc[203]=0
df_body['Month']=df_body['Month'].apply(lambda x:'May ' if x=='May & june' else x)
df_body['Month'].value_counts()
#================================ 
df_body["Israelis Injuries"]=df_body["Israelis Injuries"].astype(str)
df_body["Palestinians Injuries"]=df_body["Palestinians Injuries"].astype(str)
df_body["Palestinians Injuries"]=df_body["Palestinians Injuries"].apply(lambda x:x.replace(',',''))
df_body["Israelis Injuries"]=df_body["Israelis Injuries"].apply(lambda x:x.replace(',',''))
df_body["Israelis Injuries"]=df_body["Israelis Injuries"].astype(int)
df_body["Palestinians Injuries"]=df_body["Palestinians Injuries"].astype(int)
#=================== 
df_body["Israelis Killed"]=df_body["Israelis Killed"].astype(str)
df_body['Palestinians Killed']=df_body['Palestinians Killed'].astype(str)
df_body["Israelis Killed"]=df_body["Israelis Killed"].apply(lambda x:x.replace(',',''))
df_body['Palestinians Killed']=df_body['Palestinians Killed'].apply(lambda x:x.replace(',',''))
df_body["Israelis Killed"]=df_body["Israelis Killed"].astype(int)
df_body['Palestinians Killed']=df_body['Palestinians Killed'].astype(int)
#======= 
#extracting new features 

df_body['total_victims']=df_body['Palestinians Killed']+df_body["Israelis Killed"]+df_body["Palestinians Injuries"]+df_body["Israelis Injuries"]
df_body['pal_perce']=(df_body['Palestinians Killed']+df_body["Palestinians Injuries"])/df_body['total_victims']
df_body['isr_perce']=(df_body["Israelis Killed"]+df_body["Israelis Injuries"])/df_body['total_victims']
```

* because most of the dataset features are numerical i choose to fill it with zeros 
* there are missing values in year and mounth columns which are may2021 so we replace the zeros to define the date 
* change the year feature astype (int) to get rid of decimals 
* Make the Mounth Feature capitalized ( first letter Uppercase the rest letters in lowercase) 
* remove , from numerical values by changing the feature to string and replace ',' with nothing to delet it 
* fix the outliers values in features 
* extracting new features from the data

## Home demolitions Data 

```python
df_demo.columns
```

```python
df_demo.fillna(0)
df_demo['Year']=df_demo['Year'].fillna(0)
df_demo['Year']=df_demo['Year'].apply(lambda x:2004 if x ==0 else x )
df_demo['Year']=df_demo['Year'].astype(int)
#====
df_demo['Housing Units']=df_demo['Housing Units'].fillna(0)
df_demo['Housing Units']=df_demo['Housing Units'].astype(str)
df_demo['Housing Units']=df_demo['Housing Units'].apply(lambda x:x.replace(',',''))
df_demo['Housing Units']=df_demo['Housing Units'].astype(int)
#===
df_demo['People Left Homeless']=df_demo['People Left Homeless'].fillna(0)
df_demo['People Left Homeless']=df_demo['People Left Homeless'].astype(str)
df_demo['People Left Homeless']=df_demo['People Left Homeless'].apply(lambda x:x.replace(',',''))
df_demo['People Left Homeless']=df_demo['People Left Homeless'].astype(int)
#==== 
df_demo['Minors Left Homeless']=df_demo['Minors Left Homeless'].fillna(0)
df_demo['Minors Left Homeless']=df_demo['Minors Left Homeless'].astype(str)
df_demo['Minors Left Homeless']=df_demo['Minors Left Homeless'].apply(lambda x:x.replace(',',''))
df_demo['Minors Left Homeless']=df_demo['Minors Left Homeless'].astype(int)
```
