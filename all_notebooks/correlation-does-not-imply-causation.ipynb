# Introduction
A common mistake in statistics is to deduce a cause-and-effect relationship between two variables solely on the basis of an observed association or correlation between them.

One of the most famous examples is the Pirates versus Global Warming used as a parody by the author of the Flying Spaghetti Monster to debunk other misleading graphs. 

I his example Somalia has "the highest number of pirates and the lowest carbon emissions of any country".

```python
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
file = '../input/pirates-vs-global-warming.csv'
import pandas as pd
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource
from bokeh.models.annotations import Title
df = pd.read_csv(file)
source = ColumnDataSource(df)
plot = figure()
t = Title()
t.text = 'Global Average Temperature (celsius) vs Number of Pirates (approximate)'
plot.title=t
legend_label = 'Years (1820-2000)'
x='global-average-temperature'
y='number-of-pirates'
color='green'
plot.line(x=x, y=y, line_width=1, source=source, color=color, legend_label=legend_label)
output_notebook() # show the output in jupyter notebook
show(plot)
```
