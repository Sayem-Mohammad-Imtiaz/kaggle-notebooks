```python
import pandas as pd
```

```python
train=pd.read_csv("../input/train.csv")
```

```python
train.head()
```

```python
train.info()
```

```python
train.shape
```

```python
train["Age"].mean()
```

```python
train.drop("Cabin", axis=1,inplace=True)
```

```python
train.describe()
```

```python
train.mean()
```

```python
train.fillna(round(train.mean(),2), inplace=True)
```

```python
train.head(10)
```

```python
train.dropna(axis=0,inplace=True) 
```

```python
from sklearn.cluster import KMeans

```

```python
train_numeric=train.select_dtypes(exclude=['object'])
```

```python
train_numeric.drop("Survived",axis=1,inplace=True)
```

```python
kmeans = KMeans(n_clusters=2)
```

```python
kmeans.fit(train_numeric)
```

```python
train["cluster"]=kmeans.predict(train_numeric)
```
