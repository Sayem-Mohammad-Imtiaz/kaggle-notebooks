```python
dataset = pd.read_csv("/kaggle/input/cardekho/car data.csv")
dataset.head()
```

```python
dataset.shape
```

```python
dataset.isnull().any()
```

# We will drop the car name field because we don't predict the selling price of the car based on the model of car

```python
dataset.drop(['Car_Name'], inplace=True, axis=1)
dataset
```

### As fuel type, seller type and transmission are of categorical type value. So, we will be dividing those categorical values into sub categories

```python
print(dataset['Fuel_Type'].unique())
print(dataset['Seller_Type'].unique())
print(dataset['Transmission'].unique())
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset
```

# Now, we will be not calculating based on the year instead we will make predictions based on the number of years completed after buying the car

```python
# Now we will use datetime to calculate the number of years completed.

from datetime import date
today = date.today()
current_year = today.year
print(current_year)
```

# Now, we will replace Year with Number of Years

```python
dataset['Current_Year'] = current_year
dataset
```

```python
dataset['Number_of_Years'] = dataset['Current_Year'] - dataset['Year']
dataset
```

```python
dataset.drop(['Year', 'Current_Year'], inplace=True, axis=1)
dataset
```

# The above table is our final table and using this table we need to make predictions.

```python
X = dataset.iloc[:, 1:]
y = dataset.iloc[:, 0]
```

```python
#Dividing the dataset into X and Y
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 5)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
# Making predictions using Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
```

```python
RF_regressor = RandomForestRegressor(n_estimators = 125)
```

```python
RF_regressor.fit(X_train, y_train)
```

```python
y_pred = RF_regressor.predict(X_test)
y_pred
```

# Checking the error rate and success rate

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
print(mean_squared_error(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(r2_score(y_test, y_pred))
```
