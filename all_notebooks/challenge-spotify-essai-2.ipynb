```python
df=pd.read_csv('../input/spotify-dataset-19212020-160k-tracks/data_by_genres.csv')
df.head()
```

```python
nom_colonnes=list(df.columns)
```

On cherche toutes les musiques rap et classiques du dataframe

```python
genres=list(df.genres)
indice_lignes=[]
for i in range (len(genres)):
    if 'rap' in genres[i]:
        indice_lignes.append(i)

indice_lignes2=[]
for i in range (len(genres)):
    if 'classic' in genres[i]:
        indice_lignes2.append(i)
```

```python
tab=[]
for i in indice_lignes:
    tab.append(list(df.iloc[i]))
for i in indice_lignes2:
    tab.append(list(df.iloc[i]))
```

```python
df2=pd.DataFrame(tab,columns=nom_colonnes)
```

```python
df2
```

```python
from sklearn.cluster import KMeans
import seaborn as sns
from sklearn.preprocessing import normalize
```

```python
df3=df2.drop('genres',axis=1)
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=2,
    n_init=10,
    max_iter=300,
    random_state=42
)
```

```python
donnees=normalize(df3, axis=0, norm='max')
```

```python
kmeans.fit(donnees)
```

```python
kmeans.inertia_
```

```python
kmeans.labels_
liste_labels=list(kmeans.labels_)
```

Je créer une liste pour visualiser les deux clusters

```python
resultats=[[],[]]
for i in range (202):
    if liste_labels[i]==0:
        resultats[0].append(df2.loc[i]['genres'])
    else :
        resultats[1].append(df2.loc[i]['genres'])
print ('Groupe 1 :',resultats[0],'\n \n \n Groupe 2:',resultats[1])
```

Calcul du pourcentage de chansons bien classées

```python
nombre_total=202
nombre_total_rap=82
nombre_total_classic=120
n_vrai_classic=0
n_faux_classic=0
for genres in resultats[0]:
    if 'classic' in genres:
        n_vrai_classic+=1
    else :
        n_faux_classic+=1

pourcentage_vrai_positif=(n_vrai_classic+(nombre_total_rap-n_faux_classic))/202*100
print('Pourcentage de chansons bien classées:' ,pourcentage_vrai_positif,'%')
```
