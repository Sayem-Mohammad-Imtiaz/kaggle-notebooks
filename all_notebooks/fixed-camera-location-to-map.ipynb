```python
import folium
import json
```

## Preparing a map of Seoul

```python
# center of seoul
center = [37.541, 126.986]
# make a map
map = folium.Map(location=center, zoom_start=12)
map
```

## Mark area of district on map

```python
# get geodata of district in Seoul
smfile = open('../input/fixed-enforcement-camera-of-seoul-south-korea/seoul-map.json', 'r', encoding='utf-8').read()
districtLocData = json.loads(smfile)
folium.GeoJson(districtLocData, name='json_data').add_to(map)
```

```python
# check map
# there are 25 districts which is called 'gu'.
map
```

## Get location data of cameras and mark on map

```python
# get location data
df_camera = pd.read_csv('../input/fixed-enforcement-camera-of-seoul-south-korea/fixed_cctv_for_parking_enforcement.csv', encoding='utf-8')
df_camera.head()
```

```python
# there are too many cameras so I use data of "Gangnam-gu" only.
df_cam_gangnam = df_camera[df_camera['district'] == 'Gangnam-gu']
```

```python
# mark location on map
for i in range(0, len(df_cam_gangnam)):
  folium.Marker(
    location = [df_cam_gangnam.iloc[i]['latitude'], df_cam_gangnam.iloc[i]['longitude']],
    icon=folium.Icon(color="red")
  ).add_to(map)
```
