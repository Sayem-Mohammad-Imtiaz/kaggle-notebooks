```python
import numpy as np
import pandas as pd
```

```python
# Load sample prices data
prices = pd.read_csv('../input/python/sample_prices.csv')
```

```python
# Calculate returns and drop nan values
returns = prices.pct_change().dropna()
```

```python
#Calculate Standard Deviation (sample standard deviation)
returns.std()
```

```python
#Calculate Standar Deviation (Population standard deviation) on a manual way
#Note that is different from the last desciation.
deviations = returns - returns.mean()
squared_deviations = deviations**2
variance = squared_deviations.mean()
volatility = np.sqrt(variance)
volatility
```

```python
#Calculate Standard Deviation (sample standard deviation) on a manual way
number_of_obs = returns.shape[0]
variance = squared_deviations.sum() / (number_of_obs - 1)
volatility = variance**0.5
```

```python
#Anualize the standard deviation's return
returns.std()*np.sqrt(12)
```

## Case 1 - Analizing returns of Low and High deciles of stocks with more market cap. 

```python
returns = pd.read_csv('../input/python/Portfolios_Formed_on_ME_monthly_EW.csv',
                     header=0, index_col=0, parse_dates=True, na_values=-99.99)
returns.head()
```

```python
#Present the returns as decimals.
returns = returns/100

#Select the lowest 10% and highest 10% of market cap returns
columns = ['Lo 10', 'Hi 10']
returns = returns[columns]
returns.head()
```

```python
#Change the name of the columns of the dataframe
returns.columns = ['SmallCap','LargeCap']
```

```python
#Plot the returns
returns.plot.line()
```

```python
#Calculate Volatility
returns.std()
```

```python
#Calculate anualized volatility
annualized_vol = returns.std()*np.sqrt(12)
annualized_vol
```

```python
#Calculate the returns per month
n_months = returns.shape[0]
returns_per_month = (returns+1).prod()**(1/n_months)-1
returns_per_month
```

```python
#Annualized returns 
annualized_returns = (returns_per_month+1)**12 - 1
annualized_returns
```

```python
#Annualized returns (Short way)
annualized_returns = (returns+1).prod()**(12/n_months)-1
annualized_returns
```

```python
#Annualized returns for each point of volatility
annualized_returns/annualized_vol
```

```python
#Calculate sharpe ratio
riskfree_rate = 0.03
excess_return = annualized_returns - riskfree_rate
sharpe_ratio = excess_return / annualized_vol
sharpe_ratio
```
