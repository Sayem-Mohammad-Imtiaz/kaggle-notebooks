```python
train_data_all = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data = train_data_all.sample(frac=0.75,random_state=200) #random state is a seed value
train_data.head()
```

```python
# train_data.sort_values(by=["Age"]).style
```

```python
validation_data = train_data_all.drop(train_data.index)
validation_data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_data["Survived"]

features = ["Pclass", "Sex", "SibSp", "Parch", "Fare"]
X = pd.get_dummies(train_data[features]) # Convert categorical variable into dummy/indicator variables.
X_vldtn = pd.get_dummies(validation_data[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_vldtn)

validation = pd.DataFrame({'PassengerId': validation_data.PassengerId, 'Survived': validation_data.Survived, 'SurvivedPred': predictions})
validation["AccuratePrediction"] = validation["Survived"] == validation["SurvivedPred"]
print("accuracy %:", 100*sum(validation["AccuratePrediction"])/len(validation))
```

```python
train_data_all = pd.read_csv("/kaggle/input/titanic/train.csv")
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
# test_data.head()
```
