```python
df = pd.read_csv("/kaggle/input/iris-flower-dataset/IRIS.csv")
```

```python
set(df["species"])
```

```python
df["sp"]=df["species"].map({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})
```

```python
df.head()
```

```python
df1 = df.drop("species", axis=1)
```

```python
df1.head()
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
import sklearn
from sklearn.model_selection import train_test_split
```

```python
X = df1.drop("sp", axis=1)
y = df1["sp"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.33,random_state=32)
```

```python
X_traint = torch.FloatTensor(X_train.values)
X_testt = torch.FloatTensor(X_test.values)
y_traint = torch.LongTensor(y_train.values)
y_testt  = torch.LongTensor(y_test.values)
```

```python
class Iris(nn.Module):
    def __init__(self, inputs=4, hidden=1000, output=3):
        super().__init__()
        self.fc1 = nn.Linear(inputs,hidden)
        self.fc2 = nn.Linear(hidden,output)
        #self.fc3 = nn.Softmax(dim=1)
    def forward(self,x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        #x = self.fc3(x)
        return(x)

model = Iris()    
criterian = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

  
    
```

```python
epochs = 10000

for i in range(epochs):
    ypred = model.forward(X_traint)
    #print(ypred)
    loss = criterian(ypred,y_traint)
    print(loss)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
