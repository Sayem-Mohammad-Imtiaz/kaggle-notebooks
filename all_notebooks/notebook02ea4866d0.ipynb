```python
data = pd.read_csv('/kaggle/input/szeged-weather/weatherHistory.csv')

#handling missing values
clean_data = data.fillna(method = 'bfill' , axis = 0)
clean_data.isna().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(data['Humidity'] , data['Temperature (C)'] )
plt.show()
```

From the the data visualisation between temperature as above

From the the data visualisation between temperature (as a dependant variable) and Humidity (as an independant variable) above, we can see the form witch can fit a negative linear regression line 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
reg = LinearRegression()
X = np.array(clean_data['Humidity']).reshape(-1 , 1)
y = np.array(clean_data['Temperature (C)']).reshape(-1 , 1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
reg.fit(X_train, y_train)
print('Coefficient:', reg.coef_)
print('Intercept:', reg.intercept_)
```

```python
from sklearn.metrics import r2_score
y_predicted = reg.predict(X_test)
print('R-squared (Score):', np.sqrt(r2_score(y_test, y_predicted)))
```

As we can see the score of relaship between humidity and temperature is close to 1 than to 0. Thus, we can say that our model fits well.
