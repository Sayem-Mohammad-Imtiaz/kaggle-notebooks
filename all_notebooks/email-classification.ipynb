```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
msgs = pd.read_csv("../input/spam-and-ham/spam.csv", encoding="ISO-8859-1")
```

```python
X = msgs['v2']
y = msgs['v1']
```

```python
msgs.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from string import punctuation
from nltk.stem import PorterStemmer, WordNetLemmatizer
```

```python
st = set(stopwords.words("english"))
punc = set(punctuation)
ps = PorterStemmer()

def custom(sent):
    words = word_tokenize(sent.lower())
    faaltu = st.union(punc)
    pure = [w for w in words if w not in faaltu]
    more_pure = list(map(ps.stem, pure))
    
    return more_pure
```

```python
cv = CountVectorizer(tokenizer=custom)
```

```python
cv.fit(X)
```

```python
X_gen = cv.transform(X).todense()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
...     X_gen, y, test_size=0.33, random_state=42)
```

```python
y.shape
```

```python
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
