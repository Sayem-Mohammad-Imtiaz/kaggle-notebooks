```python
import plotly.express as px

df = px.data.stocks(indexed=True)-1

# turn index to datetime index
df.index = pd.to_datetime(df.index)
df.info()
df.head()
```

```python
df.index
```

```python
# plot Google
fig = px.line(df, x=df.index, y="GOOG")
fig.show()
```

```python
df.index.year.value_counts()
```

```python
# plot only Google data for year 2018 with range slider
df['date'] = df.index
fig = px.line(df, x='date', y='GOOG', range_x=['2018-01-01','2018-12-31'])
fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)
fig.show()
```

```python
# plot graph while hiding gaps in data
import plotly.express as px
import pandas as pd

df = pd.read_csv('../input/financestocks/stocks.csv', parse_dates=['Date'])

fig = px.scatter(df, x='Date', y='AAPL.High', range_x=['2015-12-01', '2016-01-15'],
                 title="Hide Weekend and Holiday Gaps with rangebreaks")
fig.update_xaxes(
    rangebreaks=[
        dict(bounds=["sat", "mon"]), #hide weekends
        dict(values=["2015-12-25", "2016-01-01"])  # hide Christmas and New Year's
    ]
)
fig.show()
```

```python
df.index = df['Date']
df_plot = df.drop(labels='Date', axis=1)
df_plot.head()
```

```python
import plotly.express as px

# group data by year
df = df_plot.groupby(by=df_plot.index.isocalendar().year).mean()

fig = px.bar(df, x=df.index, y='AAPL.High',
                 title="APPL High by year")

fig.update_layout(
    xaxis = dict(
        tickmode = 'linear',
        tick0 = 2015,
        dtick = 1
    )
)
fig.show()
```
