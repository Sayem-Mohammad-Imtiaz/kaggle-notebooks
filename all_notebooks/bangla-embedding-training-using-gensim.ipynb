# Importing and reading Data

```python
df = pd.read_csv('../input/bangla-wikipedia-dataset/wiki.csv')
df.head()
```

# Cleaning Sentences

```python
def remove_punc(sentences):
    new_sentences=[]
    exclude = list(set(string.punctuation))
    exclude.extend(["’", "‘", "—","\n",'০','১','২','৩','৪','৫','৬','৭','৮','৯'])
    for sentence in sentences:
        s = ''.join(ch for ch in sentence if ch not in exclude)
        new_sentences.append(s)
    
    return new_sentences
```

```python
sentence = []
for i in range(70377):
    sentence.append(df.loc[i, 'text'])
clean_sentence = remove_punc(sentence)
```

```python
clean_sentence =[article.split('।') for article in clean_sentence]
clean_sentence=[item for sublist in clean_sentence for item in sublist]
clean_sentence=[item.strip() for item in clean_sentence if len(item.split())>1]
clean_sentence=[item.split() for item in clean_sentence]
print(clean_sentence[:5])
```

# Train using Word2Vec

```python
w2v_model = Word2Vec(clean_sentence, size=300, window=5, min_count=3)
```

```python
# summarize the loaded model
print(w2v_model)
# summarize vocabulary
words = list(w2v_model.wv.vocab)
print('Number of vocabs are',len(words))
# access vector for one word
print(w2v_model['sentence'])
# save model
w2v_model.save('w2v_model.bin')

# Uncomment the lines below to load model
# new_model = Word2Vec.load('w2v_model.bin')
# print(new_model)
```

```python

print("What are the words most similar to ছেলে")
girl=w2v_model.wv.most_similar('ছেলে', topn=5)[0][0]
print(w2v_model.wv.most_similar('ছেলে', topn=5)[0][0])
```
