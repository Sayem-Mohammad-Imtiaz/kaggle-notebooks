```python
%%javascript

let boolShowNames = true;   
let btnShowHide = document.getElementById('btn_show_hide');
let names = document.querySelectorAll('.athlete text');

     btnShowHide.onclick = function(){
        boolShowNames = ! boolShowNames;

        if (boolShowNames){
            changeNamesVisibility('visible');
        }
        else{
            changeNamesVisibility('hidden');
        }
     }

     function changeNamesVisibility(visibilityCondition){
             for(let i = 0; i < names.length; i++){
                names[i].style.visibility = visibilityCondition;
             }
     }
```
