```python
###counting Fit and Occassions
to_count = ["Fit","Occasion"]
fit_dict = {}
occasion_dict = {}
for index in range(new_df.shape[0]):
    one_entry = new_df["attributes_as_dict"].iloc[index]
#     keys = one_entry.keys()

    for key,value in one_entry.items():
        value = value[0]
        key = key.strip()
        if key in to_count:
            if key==to_count[0]:
                fit_dict[value] = fit_dict.get(value,1) + 1
            else:
                occasion_dict[value] = occasion_dict.get(value,1) + 1
print("occasion_dict :",occasion_dict,"\nfit_dict:",fit_dict)
```

```python
sleeve_length = []
sleeve_occur = {}
iloc_values = []
for index,x in enumerate(new_df["attributes_as_dict"]):
    x = dict(x)
    keys = x.keys()
#     print(keys)
    if 'Sleeve Length' in keys:
        iloc_values.append(index)
        if x["Sleeve Length"] in sleeve_length:
            sleeve_occur[x["Sleeve Length"]] += 1
            continue
        sleeve_length.append(x["Sleeve Length"])
        sleeve_occur[x["Sleeve Length"]] = 1
print(sleeve_length)
print(sleeve_occur)
print(len(iloc_values))
```

```python
def get_sleeve_attribute(item):
    if "Sleeve Length" in item.keys():
        sleeve_type = item["Sleeve Length"]
        return sleeve_type
    else:
        return "No"
sleeve_df["Sleeve_length"] = sleeve_df["attributes_as_dict"].map(get_sleeve_attribute)
```

```python
sleeve_df.drop(["attributes_as_dict"],axis=1,inplace=True)
```

```python
#Let three_quarter be [ 1 , 0 , 0 , 0]
# Long Sleeves be [ 0, 1 ,0 ,0 ]
# Short Sleeves be [ 0, 0 ,1 ,0]
# Sleeveles be [ 0 , 0, 0 ,1]
# and rest be [ 0 ,0 ,0  ,0]
```

```python
sleeve_df["Sleeve_length"].value_counts()
```

```python
dict_sleeve = {
   "Three-Quarter Sleeves" : [1,0,0,0],
    "Long Sleeves" : [0,1,0,0],
    "Short Sleeves" : [0,0,1,0],
    "Sleeveless" : [0,0,0,1],
    "No" : [0,0,0,0],
    "No Sleeves": [0,0,0,0]
}
```

```python
def encoding(item):
#     print(item)
    value = dict_sleeve[item]
    return value
sleeve_df["Sleeve_length"] = sleeve_df["Sleeve_length"].map(encoding)
```
