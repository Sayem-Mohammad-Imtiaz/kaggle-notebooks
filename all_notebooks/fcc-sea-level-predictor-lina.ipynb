```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
from scipy.stats import linregress
```

```python
df = pd.read_csv('../input/sea-level/epa-sea-level.csv')
df
```

```python
#Create scatter plot
plt.figure(figsize = (12,8))
plt.scatter(df['Year'], df['CSIRO Adjusted Sea Level'])

#Create best fit line from year 1880-2050
line_a = stats.linregress(df['Year'], df['CSIRO Adjusted Sea Level'])
x = np.arange(df['Year'].iloc[0], 2050)
y = line_a.intercept + line_a.slope*x
plt.plot(x, y, color = 'green', label = "Line of best fit from 1880-2050")

#Create best fit line from year 2000-2050
df2k = df[df['Year']>=2000]
line_b = stats.linregress(df2k['Year'], df2k['CSIRO Adjusted Sea Level'])
x1 = np.arange(2000, 2050)
y1 = line_b.intercept + line_b.slope*x1
plt.plot(x1, y1, color = 'red', label = "Line of best fit from 2000-2050", linestyle = "--")

#Add labels to the table
plt.title ("Rise in Sea Level")
plt.xlabel("Year")
plt.ylabel("CSIRO Adjusted Sea Level")
plt.legend()
plt.show()
```
