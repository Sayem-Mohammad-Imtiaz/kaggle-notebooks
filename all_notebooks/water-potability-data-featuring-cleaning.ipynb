```python
(df.isnull().sum() / len(df)) * 100
```

So, we've just discovered that we only have missing data in ph, Sulfate and Trihalomethanes with ~15%, ~24% and ~5% respectively.
We will plot this just to visualization purposes.

```python
def percent_missing(df):
    percent_nan = 100* df.isnull().sum() / len(df)
    percent_nan = percent_nan[percent_nan > 0].sort_values()
    return percent_nan
```

```python
percente_data_missing = percent_missing(df)
```

```python
sns.barplot(x=percente_data_missing.index,y=percente_data_missing)
plt.xticks(rotation=90)
plt.show()
```

by now we should analize which, by the following options, is a more worth thing to do, in this situation:

1. eliminate the whole column
2. Assume NaN is equal to 0
3. fill NaN with the data average in the respective column

as long as we don't have a dataset, where is something like >>70% to make our future model a poor training, it's NOT recommended to do in this situation even because we are working with something that directly make the water potable or not.
if we assume that NaN is equal to 0 we are saying that the first row have 0 by his ph which don't make sense, chemicaly speaking, or even if we attempt to fill with 7 ( neutral ph ) still doesn't make sense because ph should change by the other characteristics we have in the other columns.
the 3 options is the most reasonable here.

```python
corr = df.corr()
```

```python
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
plt.show()
```

```python
df['ph'] = df['ph'].fillna(df['ph'].mean())
df['Trihalomethanes'] = df['Trihalomethanes'].fillna(df['Trihalomethanes'].mean())
df['Sulfate'] = df['Sulfate'].fillna(df['Sulfate'].mean())
```

```python
new_percente_data_missing = percent_missing(df)
new_percente_data_missing
```

So we no longer have missing data
