# Task 1: Best Selling Category

```python
product_group = df.groupby('category_name_1')
quantity_ordered = product_group.sum()['qty_ordered']

quantity_ordered
```

***The most ordered item***

```python
plt.figure(figsize=(10,9))
products = [product for product, df in product_group] # for mainting the consistency among data

sns.barplot(x = products, y = quantity_ordered,)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.xticks(rotation='90')
plt.show()
```

# Task 1 Answer: The most selling category is Mobiles & Tablets and second one is Men's Fashion.
