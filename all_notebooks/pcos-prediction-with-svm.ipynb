### Train Model

```python
from sklearn.svm import SVC
model2 = SVC(kernel ='linear').fit(X_train,y_train)
print("X_shape :", X.shape)
X_train = X_train[:100][:]
X_test = X_test[:100][:]
y_train = y_train[:100][:]
y_test = y_test[:100][:]

#dự đoán trên tập test
y_pred2 = model2.predict(X_test)
```
