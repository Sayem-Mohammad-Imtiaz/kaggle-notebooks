```python
# I am a new to machine learning, so the code won't be the best but I've tried my best to solve this problem. 
# If any of y'all are experts, your feedback would be of grate help 

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import time # This is not really needed

# imported the data
data = pd.read_csv("/kaggle/input/breast-cancer-wisconsin-data/data.csv")

# Split the data into the labels and targets (1 are the labels, 2 is the target)
X1 = data.drop(data.columns.difference(['radius_mean','perimeter_mean']), 1)
X2 = data['diagnosis']

y1 = data.drop(data.columns.difference(['texture_mean','area_mean']), 1)
y2 = data['diagnosis']

z1 = data.drop(data.columns.difference(['smoothness_mean', 'compactness_mean']), 1)
z2 = data['diagnosis']

# imported the model that I'm going to use (I've use decision tree because it's a classification problem)
model = DecisionTreeClassifier()

# Split the data into training and testing
X1_train, X1_test, X2_train, X2_test = train_test_split(X1,X2, test_size=0.3)
y1_train, y1_test, y2_train, y2_test = train_test_split(y1,y2, test_size=0.3)
z1_train, z1_test, z2_train, z2_test = train_test_split(z1,z2, test_size=0.3)

# 
print('Calculating..........')
time.sleep(1.2)

# User inputs
radius = input("Enter radius ")
perimeter = input("Enter perimeter ")
texture = input("Enter texture ")
area = input("Enter area ")
smoothness = input("Enter smoothness ")
compactness = input("Enter compactness ")

print('Calculating..............')
time.sleep(1.2)

# Trained the model
model.fit(X1_train, X2_train)
model.fit(y1_train, y2_train)
model.fit(z1_train, z2_train)

# Used the model to make a prediction based on the inputs given by the user
pred_1 = model.predict([[(radius), (perimeter)]])
pred_2 = model.predict([[(texture), (area)]])
pred_3 = model.predict([[(smoothness), (compactness)]])

# Showing the results in 
if pred_1 and pred_2 and pred_3 == 'B':
    print('[B]')
elif pred_1 and pred_2 and pred_3 == 'M':
    print('[M]')
else:
    print(pred_1)
    print(pred_2)
    print(pred_3)
```
