```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading data

```python
# Read data
all_data = pd.read_csv("../input/cvdcvd-vd/Social_Network_Ads.csv")
all_data["Gender"] = pd.factorize(all_data["Gender"])[0]
all_data.head()
```

```python
all_data.describe()
```

# Feature analysis and visualization

```python
sns.pairplot(all_data, hue="Purchased")
```

```python
corr_matrix = all_data.corr()
sns.heatmap(corr_matrix, annot=True)
```

# Model

```python
# Import model
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
columns = ["Age", "EstimatedSalary"]
X = all_data[columns].values
y = all_data["Purchased"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50, random_state=4)
```

```python
# Train model
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# Evaluate model
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_pred, y_test)
print(f"Model accuracy: {accuracy}")
```
