```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path_train="../input/car-price-prediction/train.csv"
data= pd.read_csv(path_train)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

```python
cat_cols = [cname for cname in data.columns if
                    data[cname].nunique() < 10 and 
                    data[cname].dtype == "object"]

cat_transformer = Pipeline(steps=[
    ('onehot',OneHotEncoder(handle_unknown='ignore'))
])
preproces = ColumnTransformer(transformers=[
    ("cat",cat_transformer,cat_cols)
])
```

```python
Y= data['Price']
X= data.drop(['Price'], axis=1)
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.0001, random_state=10)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression(normalize=True)
```

```python
pipe_model = Pipeline(steps=[
    ("preprocessing",preproces),
    ('model', model)
])
```

```python
pipe_model.fit(X_train,Y_train)
```

```python
preds = pipe_model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
score = mean_squared_error(Y_test, preds)
print('MAE:', score)
```
