```python
#Code by Aristotle609 https://www.kaggle.com/aristotle609/eda-on-medium-search-dataset/notebook

def plot_maker(cat_feature):
    plot = alt.Chart(df).mark_bar().encode(
        y = cat_feature,
        x = "count()"
    ).interactive()
    return plot
```
