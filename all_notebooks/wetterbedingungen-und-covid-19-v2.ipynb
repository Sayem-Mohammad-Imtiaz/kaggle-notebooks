# Hinführung zum Thema

COVID-19 ist eine neue und unbekannte Virus-Krankheit. Alle Influenzaviren verbreiten sich rasend schnell in der Gesellschaft [1]. Unsere Forschung befasst sich mit der Ananlyse von Daten und Erstellung von Modellen, um unsere Arbeitshypothese zu beantworten.

Dass die Ausbreitung von herkömmlichen Influenzaviren mit der Temperatur und der relativen Luftfeuchtigkeit zusammenhängt, ist bewiesen [2]. Der Einfluss der Temperatur und der relativen Luftfeuchtigkeit auf Covid-19 wird aktuell noch untersucht [3].

Coronaviren gehören zur "Familie" der Viren, die bei Tieren oder Menschen Krankheiten verursachen können. Es ist bekannt, dass beim Menschen mehrere Coronaviren Infektionen der Atemwege verursachen, die von Erkältungen bis zu schwereren Erkrankungen wie dem Middle East Respiratory Syndrome (MERS) und dem Severe Acute Respiratory Syndrome (SARS) reichen. Das zuletzt entdeckte Coronavirus verursacht die Coronavirus-Krankheit COVID-19 [4]. Das Virus wird mit Sars-CoV-2 bezeichnet, die dadurch ausgelöste Krankheit wird als Covid-19 bezeichnet [5].

# Übertragungswege

### Infektion durch Tröpfchen

Nach aktuellen Studien ist die Tröpfcheninfektion der Hauptübertragungsweg von Sars-Cov-2. Diese Tröpfchen entstehen beispielsweise beim Niesen, oder beim Husten und werden von anderen Personen über die Schleimhäute, Nase oder den Mund aufgenommen [6].

### Aerosole

Aerosole sind Tröpfchenkerne, die kleiner als 5 Mikrometer sind. Bei mehreren Studien wurde aufgezeigt, dass sich in der Ausatmeluft, oder in Zimmern von Betroffenen Sars-Cov-2-Viren gibt[7-9].Durch das Tragen von chirugischen Masken (FFP3) kann eine Ausbreitung von Sars-Cov-2 haltigen Aerosolen verhindert werden[10]. Durch lautes Sprechen können ebenfalls Aerosole entstehen, die das Virus übertragen können.[11].

### Übertragung durch Kontakt

Die Übertragung des Virus über kontaminierte Oberflächen ist nicht auszuschließen, vorallem nicht in unmittelbarer Nähe von bereits infizierten Personen. Sars-Cov-2 Viren können in der Umwelt und auf Oberfllächen nachgewisen werden[12].Auch im Stuhl von Covid-19 Infizierten kann das Virus postitiv nachgewisen werden[13].

### Übertragung während der Schwangerschaft

Zum aktuellem Zeitpunkt kann eine Übertragung des Erregers von einer Infizierten Mutter auf das ungeborene Kind nicht ausgeschlossen werden[14]. Jedoch sind bei den aller meisten Fällen die Kinder bei der Geburt negativ auf den Erreger und zeigen keinerlei Krankheitsanzeichen[15]. Bei einem postivien Fall findet die Übertragung über Tröpfcheninfektionen statt[16].

### Gesundheitssektor

Der Gesundheitssektor besitzt ein hohes Infektionsrisiko, da es hier viele Übertragungsmöglichkeiten gibt. Vorallem in disem Sektor müssen die Maßnahmen zum Eigenschutz sichergestellt werden. Das höchste Risiko sind Vorgänge bei denen Aerosole ausgestoßen werden, wie z.B. zahnärztliche Untersuchungen, weil hier viel Speichel ausgestoßen wird. Oder auch Intubationen im Krankenhaus, weil auch hier viel im Mund und Rachenbereich gearbeitet wird. Die arbeitenden Personen schützen sich mit spezillen Atemschutzmasken, um eine Gefahr, mit der Ansteckung so gering wie möglich zu halten [17].
