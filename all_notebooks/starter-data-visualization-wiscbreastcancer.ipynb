**process data: divide data into set y(label) and x(rest without 'id' and 'unamed 32')**

```python
y = data['diagnosis']
```

```python
y
```

```python
x = data.drop(columns=['id', 'diagnosis', 'Unnamed: 32'])
```

```python
x
```

**process data, standardization**

```python
#standardization
x_ = (x - x.mean()) / x.std()
```

```python
x_.describe()
#mean near zero, std is one
```

# visualization

violin plots

 use [melt()](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.melt.html?highlight=melt#pandas.DataFrame.melt) to [transform from wide format to long format ](https://en.wikipedia.org/wiki/Wide_and_narrow_data)  
 because parameter 'data' of [seaborn violin plot](https://seaborn.pydata.org/generated/seaborn.violinplot.html?highlight=violinplot#seaborn.violinplot) :Dataset for plotting. If x and y are absent, this is interpreted as **wide-form**. Otherwise it is expected to be **long-form**.

```python
mean_data = pd.concat([y,x_.iloc[:,0:10]],axis=1)
mean_data = pd.melt(mean_data,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
```

```python
mean_data
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value", hue="diagnosis", data=mean_data,split=True, inner="quart")
plt.xticks(rotation=90)
```

```python
se_data = pd.concat([y,x_.iloc[:,10:20]],axis=1)
se_data = pd.melt(se_data,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value", hue="diagnosis", data=se_data,split=True, inner="quart")
plt.xticks(rotation=90)
```

```python
worst_data = pd.concat([y,x_.iloc[:,20:30]],axis=1)
worst_data = pd.melt(worst_data,id_vars="diagnosis",
                    var_name="features",
                    value_name='value')
plt.figure(figsize=(10,10))
sns.violinplot(x="features", y="value", hue="diagnosis", data=worst_data,split=True, inner="quart")
plt.xticks(rotation=90)
```
