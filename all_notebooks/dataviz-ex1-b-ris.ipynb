### Item C - Representação Gráfica
Para cada uma das variáveis, produza um ou mais gráficos, usando matplotlib ou Pandas plot, que descreva seu comportamento / característica. Lembre-se que o(s) gráfico(s) precisa(m) ser compatível(eis) com a classificação da variável. 

#### - Total Occurrences with Fatalities by Airplane Type:

```python
# Count of Airplane Type:
count_tipo = anv.groupby('Tipo')['Fatalidades'].sum().sort_values(ascending=False)

# Setting the figure size:
fig, ax = plt.subplots(figsize=(15,5))

# Setting axes lables text:
ax.set_ylabel('Total Occurrences', fontsize = '15')
ax.set_xticklabels(count_tipo.index, fontsize = '10')
plt.xticks(rotation=45)

# Title:
plt.title('Total Occurrences with Fatalities by Airplane Type', fontweight='bold',
          color = 'red', fontsize='20', horizontalalignment='center')

# Bar plot:
plt.grid()
plt.bar(count_tipo.index, count_tipo.values);
```

#### - Total Occurrences with Fatalities by Operation Phase

```python
# Count of Phase 
count_phase = anv.groupby('Fase_Operacao')['Fatalidades'].sum().sort_values(ascending=False)

# Setting the figure size:
fig, ax = plt.subplots(figsize=(15,5))

# Setting axes lables text:
ax.set_ylabel('Total Occurrences', fontsize = '15')
ax.set_xticklabels(count_phase.index, fontsize = '10')
plt.xticks(rotation=90)

# Title:
plt.title('Total Occurrences with Fatalities by Operation Phase', fontweight='bold',
          color = 'red', fontsize='20', horizontalalignment='center')

# Bar plot:
plt.grid()
plt.bar(count_phase.index[:15], count_phase.values[:15]);
```
