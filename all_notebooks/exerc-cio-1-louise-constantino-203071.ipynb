```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import squarify


df = pd.read_csv('../input/dataviz-facens-20182-aula-1-exerccio-2/anv.csv', delimiter=',')

df.head(5) 
```

```python
variaveis = [["aeronave_tipo_veiculo",    "Qualitativa Nominal"],
            ["aeronave_fabricante",      "Qualitativa Nominal"],
            ["aeronave_ano_fabricacao",  "Quantitativa Discreta"],
            ["aeronave_pais_fabricante", "Qualitativa Nominal"],            
            ["aeronave_fase_operacao",   "Qualitativa Nominal"],
            ["aeronave_nivel_dano",      "Qualitativa Nominal"],
            ["total_fatalidades",        "Quantitativa Discreta"]
           ] 

variaveis = pd.DataFrame(variaveis, columns=["Variável", "Classificação"])
variaveis
```

### Questão 1 - Item B - Tabela de frequência

### Questão 1 - Item C - Representação Gráfica
Para cada uma das variáveis, produza um ou mais gráficos, usando matplotlib, que descreva seu comportamento / caracteristica. Lembre-se que estes gráficos precisam ser compatíveis com a classificação da variável. 

```python
comMorte = df[df.total_fatalidades > 0].count()[1]
semMorte = df[df.total_fatalidades == 0].count()[1]   

labels = 'Com Morte', 'Sem Morte'
sizes = comMorte, semMorte
colors = ['lightcoral', 'lightgrey']
explode = (0, 0.1)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%', shadow=True, startangle=0)
ax1.axis('equal')
fig1.set_size_inches(6,6) 

plt.title('% Acidentes Com Mortes e Sem Mortes', fontsize=15)
plt.show()
```

```python
fat_fabr = df[['aeronave_tipo_veiculo','total_fatalidades']].groupby(['aeronave_tipo_veiculo']).sum() 
fat_fabr = fat_fabr.sort_values(by=['total_fatalidades'], ascending=True)

labels = fat_fabr.index
values = fat_fabr.total_fatalidades 
x_pos  = [i for i, _ in enumerate(labels)]

plt.figure(figsize=(20, 10))
plt.barh(x_pos, values, color='#b60000')

plt.title("Fatalidades Tipo de Veículo", fontsize=20)

plt.yticks(x_pos, labels)

plt.show()
```

```python
fat_fabr = df[df['aeronave_fabricante']!='***'][['aeronave_fabricante','total_fatalidades']].groupby(['aeronave_fabricante']).sum() 
fat_fabr = fat_fabr.sort_values(by=['total_fatalidades'], ascending=False).head(10)

labels = fat_fabr.index
values = fat_fabr.total_fatalidades 

plt.figure(figsize=(20, 7))
plt.bar(labels, values, color='#b60000')
plt.xticks(rotation=90)
plt.title("Top 10 Fabricantes com maior nº de Fatalidades", fontsize=20)

plt.show()
```

```python
pais_comMorte = df[df['total_fatalidades'] >  0][['aeronave_pais_fabricante','total_fatalidades']].groupby(['aeronave_pais_fabricante']).count() 
pais_semMorte = df[df['total_fatalidades'] == 0][['aeronave_pais_fabricante','codigo_ocorrencia']].groupby(['aeronave_pais_fabricante']).count() 

fat_paises = pd.concat([pais_comMorte, pais_semMorte], axis=1, sort=False)
fat_paises.rename(columns={'total_fatalidades': 'Com Morte'}, inplace=True)
fat_paises.rename(columns={'codigo_ocorrencia': 'Sem Morte'}, inplace=True)

fat_paises = fat_paises.sort_values(by=['Sem Morte'], ascending=False).head(5)
 

labels   = fat_paises.index
comMorte = fat_paises['Com Morte']
semMorte = fat_paises['Sem Morte']

width = 0.5

fig, ax = plt.subplots()

ax.bar(labels, comMorte, width, label='Com Morte', color='#b60000')
ax.bar(labels, semMorte, width, bottom=comMorte, label='Sem Morte', color='#AAAAAA')
ax.legend()

fig.set_size_inches(20, 10) 
plt.xticks(rotation=90)

plt.title("Top 5 - Acidentes Com Mortes e Sem Mortes por País Fabricante", fontsize=20)

plt.show()
```

```python
fat_fase = df[['aeronave_fase_operacao','total_fatalidades']].groupby(['aeronave_fase_operacao']).sum() 
fat_fase = fat_fase.sort_values(by=['total_fatalidades'], ascending=False).head(15)

labels = fat_fase.index
volume = fat_fase.total_fatalidades 

plt.figure(figsize=(20, 10))      
squarify.plot(sizes=volume, label=labels,alpha=0.6, text_kwargs={'fontsize': 10})
plt.axis('off')

plt.title('Fatalidades por Fase de Operação', fontsize=20)
plt.show()
```

```python
fat_dano = df[['aeronave_nivel_dano','total_fatalidades']].groupby(['aeronave_nivel_dano']).sum() 
fat_dano = fat_dano.sort_values(by=['total_fatalidades'], ascending=False)
labels = fat_dano.index
sizes = fat_dano.total_fatalidades 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.0f%%', shadow=True, startangle=0)
ax1.axis('equal')
fig1.set_size_inches(8,8) 

plt.title('% Fatalidades por Nível de Dano', fontsize=15)
plt.show()
```

```python
fat_ano = df[['aeronave_ano_fabricacao','total_fatalidades']].groupby(['aeronave_ano_fabricacao']).sum()
fat_ano.index = fat_ano.index.astype(int)
fat_ano = fat_ano[fat_ano.index > 0]
fat_ano.index = fat_ano.index.astype(str)
labels = fat_ano.index
values = fat_ano.total_fatalidades 
 
plt.figure(figsize=(20, 7))  
plt.fill_between( labels, values, color="lightcoral", alpha=0.2)
plt.plot(labels, values, color="red", alpha=0.5)
plt.title('% Fatalidades por Ano de Fabricação', fontsize=15)
plt.xticks(rotation=90)
plt.show()
```
