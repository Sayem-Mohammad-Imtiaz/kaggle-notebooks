# 7. Використати правило 68-95-99.7 для набору даних, визначити чи розподіл даних є подібним до нормального розподілу. Для цього вирахувати:
* Pr( µ-1σ ≤ X ≤ µ+1σ)
* Pr( µ-2σ ≤ X ≤ µ+2σ)
* Pr( µ-3σ ≤ X ≤ µ+3σ)

```python
n = PCA_var.size // 7
titles = ['PC1','PC2','PC3','PC4','PC5', 'PC6']
for i in titles:
    mu = PCA_var[f'{i}'].mean()
    sigma = PCA_var[f'{i}'].std()
    occurence = [0, 0, 0]
    for j in PCA_var[f'{i}'].values:
        if j > mu - sigma and j < mu + sigma:
            occurence[0] += 1
        if j > mu - 2 * sigma and j < mu + 2 * sigma:
            occurence[1] += 1
        if j > mu - 3 * sigma and j < mu + 3 * sigma:    
            occurence[2] += 1
    print("")
    print(f'Pr⁡(µ - 1σ ≤ {i} ≤ µ + 1σ) =', occurence[0] / n)
    print(f'Pr⁡(µ - 2σ ≤ {i} ≤ µ + 2σ) =', occurence[1] / n)
    print(f'Pr⁡(µ - 3σ ≤ {i} ≤ µ + 3σ) =', occurence[2] / n)
```

# 8.Видалити з набору даних всі дані, що не належать до інтервалу µ-3σ; µ+3σ

```python
PCA_var_cpy = PCA_var
for i in titles :
    mu = PCA_var_cpy[f'{i}'].mean()
    sigma = PCA_var_cpy[f'{i}'].std()
    irr = PCA_var[(PCA_var[f'{i}'] > mu + 3*sigma) | (PCA_var[f'{i}'] < mu - 3*sigma)].index
    PCA_var = PCA_var.drop(irr)
```

```python
print(PCA_var.size // 7)
```

**9. Провести тест Колмогорова-Смірнова.**

```python
alpha = 0.05
CB = (1 / np.sqrt(PCA_var.size // 7)) * np.sqrt(-np.log(alpha / 2))
for i in titles :
    print(f'{i}: {kstest(PCA_var[i], "norm").statistic}')

print("Critical boundary:", CB)
```

Гіпотезі про нормальний розпділ жодна з ознак на рівні значущості 0,05 не відповідає.

# Реалізувати та навчити різні моделі з різною архітектурою для розв’язання поставленої задачі регресії.

```python
from sklearn.linear_model import Lasso, Ridge, ElasticNet, LogisticRegression
from sklearn.neural_network import MLPClassifier, MLPRegressor
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeRegressor, DecisionTreeClassifier
from sklearn.ensemble import VotingRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
print(np.any(np.isnan(PCA_var)))
print(np.all(np.isfinite(PCA_var)))
PCA_var = PCA_var.dropna()
print(np.any(np.isnan(PCA_var)))
print(np.all(np.isfinite(PCA_var)))
```

```python
PCA_var.head()
```

```python
print(PCA_var.size // 7)
data0 = PCA_var.drop(columns = "Survived")
data1 = PCA_var["Survived"]
X_train, X_test, Y_train, Y_test = train_test_split(data0, data1, test_size = 0.5, random_state =0)
```

```python
regressors = (Lasso(), Ridge(), ElasticNet(), LogisticRegression(), MLPRegressor(), KNeighborsRegressor(), RandomForestRegressor(), AdaBoostRegressor(), DecisionTreeRegressor())
reg_mse_test = []
reg_mse_train = []
reg_mod_instances = []
for model in regressors :
    print(model)
    m = model
    m.fit(X_train,Y_train)
    predY_train=m.predict(X_train)
    predY_test=m.predict(X_test)
    reg_mse_train.append(mean_squared_error(Y_train,predY_train))
    reg_mse_test.append(mean_squared_error(Y_test,predY_test))
    reg_mod_instances.append(m)
```

```python
from tabulate import tabulate
l = []
for i in range(len(regressors)) :
    l.append([regressors[i], reg_mse_train[i], reg_mse_test[i]])
table = tabulate(l, headers = ['Model', 'train error', 'test error'])
print(table)
```

```python
classifiers = (MLPClassifier(), KNeighborsClassifier(), RandomForestClassifier(), AdaBoostClassifier(), DecisionTreeClassifier())
clas_mse_test = []
clas_mse_train = []
clas_mod_instances = []
for model in classifiers :
    print(model)
    m = model
    m.fit(X_train,Y_train)
    predY_train=m.predict(X_train)
    predY_test=m.predict(X_test)
    clas_mse_train.append(mean_squared_error(Y_train,predY_train))
    clas_mse_test.append(mean_squared_error(Y_test,predY_test))
    clas_mod_instances.append(m)
```

```python
l = []
for i in range(len(classifiers)) :
    l.append([classifiers[i], clas_mse_train[i], clas_mse_test[i]])
table = tabulate(l, headers = ['Model', 'train error', 'test error'])
print(table)
```

```python
voting=VotingRegressor([('Lasso',reg_mod_instances[0]),
                        ('Ridge' , reg_mod_instances[1]),
                        ('AdaBoostRegressor',reg_mod_instances[7])])
voting.fit(X_train,Y_train)
Y_predict=voting.predict(X_test)
mean_squared_error(Y_test,Y_predict)
```
