```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
%matplotlib inline

#adjust the configuration of the plot we will create
matplotlib.rcParams['figure.figsize']= (12,8)

```

```python
#read the data
file="../input/movies/movies.csv"
df = pd.read_csv(file ,encoding='ISO-8859-1')
```

```python
df.head(10)
```

```python
#check each column for missing data
for x in df.columns:
    missing= np.mean(df[x].isnull())
    print('{}-{}%'.format(x,missing))
```

```python
#data type of each column
df.dtypes
```

```python
#change the data type of specific columns
df['budget']= df['budget'].astype('int64')
df['gross']= df['gross'].astype('int64')
```

```python
df.head()
```

```python
'''create the correct year column 
by extracting the first 4 characters of released column'''

df['correct_year']= df['released'].astype(str).str[0:4]
df
```

```python
#scroll inside your dataset
pd.set_option('display.max_rows', None)
```

```python
#sort values descending by gross
df= df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#duplicates (watch unique values or drop them)
#df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#scatter plot between budget and gross with matplotlib.pyplot
#in x label you insert the dependent variable
plt.scatter(x=df['gross'], y=df['budget'])
plt.title('budget and gross correlation')
plt.xlabel('gross')
plt.ylabel('budget')
plt.show()
```

```python
#scatter plot between budget and gross with seaborn
sns.regplot(x='gross', y='budget', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
'''correlation matrix table (only on numerical fields).
you can use kendall or spearman methods'''

df.corr(method='pearson')
```

```python
matrix= df.corr(method='pearson')
sns.heatmap(matrix, annot=True)
plt.title('correlation matrix graph')
plt.xlabel('movie features')
plt.ylabel('movie features')
plt.show()
```

```python
#change object data types to category
df2=df

for x in df2.columns:
    if df2[x].dtype=='object':
        df2[x]= df2[x].astype('category')
        df2[x]= df2[x].cat.codes   #give a serial number to category columns
        
df2.head()
```

```python
#to see the original data, run the second line of code first
df.head()
```

```python
#correlation matrix for all columns after categorizing objects
matrix= df2.corr(method='pearson')
sns.heatmap(matrix, annot=True)
plt.title('correlation matrix graph')
plt.xlabel('movie features')
plt.ylabel('movie features')
plt.show()
```
