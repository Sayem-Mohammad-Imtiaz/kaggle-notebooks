```python
import pandas as pd
data=pd.read_csv("../input/star-dataset/6 class csv.csv")
```

```python
data.head()
```

```python
data.isnull().sum()
data.shape
```

```python
w=data.iloc[:,5]
z=data.iloc[:,6]
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
wt=le.fit_transform(w)
zt=le.fit_transform(z)
data["Tcolour"]=wt
data["Tspectral class"]=zt
```

```python
print(data)
```

```python
x=data[['Temperature (K)','Luminosity(L/Lo)','Radius(R/Ro)','Absolute magnitude(Mv)','Tcolour','Tspectral class']]
y=data[['Star type']]
```

```python
import seaborn as sb
sb.pairplot(data, hue="Star type")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.1,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
sc.fit(x_train)
sc.fit(x_test)
x_train=sc.transform(x_train)
x_test=sc.transform(x_test)
```

```python
x_train.shape
x_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion="entropy",random_state=0)
dt.fit(x_train,y_train)
```

```python
y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
accuracy=accuracy_score(y_test,y_pred)*100
print("accuracy of the model is "+ str(accuracy) + "%")
cm=confusion_matrix(y_test,y_pred)
print(cm)
```
