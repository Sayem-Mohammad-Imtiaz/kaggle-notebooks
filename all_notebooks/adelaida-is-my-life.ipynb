```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
astat = pd.read_csv("../input/aflstats/afl_stats.csv")
astat.info()
astat.head()
```

```python
astat1= astat.query('Team == "Adelaide"').query('Day == "Sat"')


count_l = astat1.groupby('WinLoss').count()
count_l['WL'] = ['L', "W"]
count_l

fig, ax = plt.subplots(figsize=(16, 8))

ax.barh(count_l.WL,count_l.Year, color='#189AB4', edgecolor='black')

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.set_axisbelow(True)

plt.legend(['WinLoss'])

plt.grid(alpha=0.3, axis='x')
plt.suptitle(f"Adelaide Wins and Loses on Saturday", ha='left', x='0.125', y=0.96, fontsize=18, fontweight='bold')
plt.show()
```
