```python
X_PCA_verified = PCA(n_components = 2).fit_transform(X)
PCA_table_verified = pd.DataFrame(data = X_PCA_verified, columns = ['Dimension 1', 'Dimension 2'])
PCA_table_verified['Price'] = table['price']
```

```python
plt.figure(figsize = (15, 10))
sns.heatmap(PCA_table_verified.corr(), annot = True)
```

```python
columns_names = ['Dimension 1', 'Dimension 2', 'Price']

for i in columns_names:
    m = PCA_table_verified[f'{i}'].mean()
    s = PCA_table_verified[f'{i}'].std()
    val = [0, 0, 0]
    for j in PCA_table_verified[f'{i}'].values:
        if j > m - s and j < m + s:
            val[0] += 1
        if j > m - 2 * s and j < m + 2 * s:
            val[1] += 1
        if j > m - 3 * s and j < m + 3 * s:    
            val[2] += 1
    print(f'\nPr⁡(µ - 1σ ≤ {i} ≤ µ + 1σ) =', val[0] / PCA_table_verified.shape[0])
    print(f'Pr⁡(µ - 2σ ≤ {i} ≤ µ + 2σ) =', val[1] / PCA_table_verified.shape[0])
    print(f'Pr⁡(µ - 3σ ≤ {i} ≤ µ + 3σ) =', val[2] / PCA_table_verified.shape[0])

print('\n')
```

```python
def resultOfTest(statistic):
    critical_value = (1 / np.sqrt( 50)) * np.sqrt(-np.log(0.05 / 2))
    if statistic < critical_value:
        return f'Sample has a normal distribution, α = 0.05'
    else:
        return f'Sample does not have a normal distribution, α = 0.05'
```

```python
result = kstest(PCA_table_verified['Dimension 1'], 'norm').statistic
print('Result =', result)
print(resultOfTest(result))
```

```python
result = kstest(PCA_table_verified['Dimension 2'], 'norm').statistic
print('Result =', result)
print(resultOfTest(result))
```

```python
result = kstest(PCA_table_verified['Price'], 'norm').statistic
print('Result =', result)
print(resultOfTest(result))
```

```python
X = PCA_table_verified.loc[:,('Dimension 1', 'Dimension 2')].values
Y = PCA_table_verified['Price']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.5)
```

```python
Lasso = LassoCV()
lasso = Lasso.fit(X_train, Y_train)
Y_pred_lasso = lasso.predict(X_test)
print('Predicted price using lasso:', Y_pred_lasso)
```

```python
accuracy_lasso = r2_score(Y_test, Y_pred_lasso)
print('Accuracy: ', accuracy_lasso)
```

```python
train_err = 1 - lasso.score(X_train, Y_train)
test_err = 1 - lasso.score(X_test, Y_test)

print('Train error =', train_err)
print('Test error =', test_err)
print('Root-mean-square deviation =', mean_squared_error(Y_test, Y_pred_lasso, squared = False))
```
