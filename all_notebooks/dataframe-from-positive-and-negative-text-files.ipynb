```python
def parse_my_file(filename):
    with open(filename) as f:
        for line in f:
            yield line.strip().split('\n', 1)
```

```python
df = pd.DataFrame(parse_my_file('../input/ts-naim-mhedhbi/positive_tweets.txt'))
df.shape
```

```python
df = df.rename(columns={0:'texts'})
df.columns = ['texts']
df
```

```python
data_labels = [] 
for row in df['texts'] :
  data_labels.append('1') 
    
        
df['data_labels'] = data_labels
```

```python
df
```

```python
df1 = pd.DataFrame(parse_my_file('../input/ts-naim-mhedhbi/negative_tweets.txt'))
df1.shape
```

```python
df1 = df1.rename(columns={0:'texts'})
df1.columns = ['texts']
df1
```

```python
data_labels = [] 
for row in df1['texts'] :
  data_labels.append('0') 
    
        
df1['data_labels'] = data_labels
df1
```

```python
df_tun = pd.concat([df,df1], ignore_index=True)
df_tun
```
