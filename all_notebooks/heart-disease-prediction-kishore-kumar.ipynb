```python
HeartDis_data['BPMeds'] = HeartDis_data['BPMeds'].fillna(HeartDis_data['BPMeds'].median())
HeartDis_data['totChol'] = HeartDis_data['totChol'].fillna(HeartDis_data['totChol'].median())
HeartDis_data['BMI'] = HeartDis_data['BMI'].fillna(HeartDis_data['BMI'].median())
HeartDis_data['heartRate'] = HeartDis_data['heartRate'].fillna(HeartDis_data['heartRate'].median())
HeartDis_data['glucose'] = HeartDis_data['glucose'].fillna(HeartDis_data['glucose'].median())
HeartDis_data1=HeartDis_data.dropna()
```

```python
HeartDis_data1.isnull().sum()
```

```python
HeartDis_data1
```

```python
x=HeartDis_data1.iloc[:,0:-2].values
y=HeartDis_data1.iloc[:,-1].values
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)*100
```
