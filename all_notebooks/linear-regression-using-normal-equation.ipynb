```python
import pandas as pd
import numpy as np

data_frame = pd.read_csv('../input/linear-regression-dataset/Linear Regression - Sheet1.csv') # https://www.kaggle.com/tanuprabhu/linear-regression-dataset

# Normal Equation :  ( X^T.X )^-1 . (X^T.y)

def multiplication(metrixA, metrixB): # Multiply 2 metrices
  resultant_metrix = [[0 for i in range(len(metrixB[0]))] for j in range(len(metrixA))]
  for i in range(len(metrixA)):
    for j in range(len(metrixB[0])):
      for k in range(len(metrixB)):
        resultant_metrix[i][j] += metrixA[i][k] * metrixB[k][j]
  return resultant_metrix

def inverse(metrix): # Inverse of metrix
  return np.linalg.inv(metrix)

def transpose(metrix): # Transpose of metrix
  resultant_metrix = [[0 for j in range(len(metrix))] for i in range(len(metrix[0]))]
  for i in range(len(metrix)):
    for j in range(len(metrix[0])):
      resultant_metrix[j][i] = metrix[i][j]
  return resultant_metrix

def predict(x, theta0, theta1): # Predict
  return theta0 + theta1 * x

def normal_equation(x, y): # Normal Equation
  first_half = inverse(multiplication(transpose(x), x)) # ( X^T.X )^-1
  second_half = multiplication(transpose(x), y) # (X^T.y)
  return multiplication(first_half, second_half)

def main():
  x_train = data_frame.iloc[:, 0][:280] # Taking 280 columns out of 300 | 20 used for testing and prediction
  y_train = data_frame.iloc[:, 1][:280] # Taking 280 columns out of 300 | 20 used for testing and prediction
  x_metrix = []
  y_metrix = []
  for each in x_train: # Generating X
   x_metrix.append([1, each])
  for each in y_train: # Generating y
   y_metrix.append([each])
  final_metrix = normal_equation(x_metrix, y_metrix) # Calculate normal equation
  theta0 = final_metrix[0][0] # Extract theta0 from reultant metrix
  theta1 = final_metrix[1][0] # Extract theta1 from reultant metrix
  print('Prediction: ' + str(predict(280, theta0, theta1)))  

main()
```
