```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

import chardet
np.random.seed(0)

import os
print(os.listdir("../input/character-encoding-examples"))

def encoding_to_usf8(name):  
    print('+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+-+')
    file_name = '../input/character-encoding-examples/'+name
    
    #Determine encoding
    with open(file_name, 'rb') as fb:
        encod = chardet.detect(fb.read(100000))
    if (encod != 'utf-8'):
        with open(file_name, 'r', encoding=encod['encoding']) as f:
            text = f.read()
            print(text[:200])
        
    #Write text to new file with UTF-8 encoding (by default) 
    new_name = file_name
    if(name.find('.csv')==-1):
        new_name = name[:name.find('_')]+'_UTF-8.txt'
        print(name)
        print(new_name)
        with open(new_name, 'w') as fw:
            fw.write(text)
    #Make sure that is right encoding
    with open(new_name, 'rb') as fb_new:
        encod_new = chardet.detect(fb_new.read(100000))
        print(encod_new)
        
    with open(new_name, 'r') as f_new:
        text_new = f_new.read()
        print(text_new[:200])    
    return new_name


for file in os.listdir("../input/character-encoding-examples"):
    encoding_to_usf8(file)
```
