```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
wineDB = pd.read_csv('../input/winequalityred/winequality-red.csv')
```

```python
wineDB.head()
wineDB.describe().transpose()
```

```python
print("Features included in Wine dataset: ",wineDB.keys().values)
#all data are numeric
print("Target: ",wineDB.keys()[11])
```

```python
plt.figure(figsize=(20,16))
plt.title("Heatmap to visualize the data")
sns.heatmap(wineDB.corr(), annot=True)
plt.show()
```

```python

fig = plt.figure()
plt.title("scatter plot to visualize some features")
ax = plt.axes()
ax.scatter(wineDB['chlorides'], wineDB['quality'], marker='x', color='red');
plt.xlabel('chlorides')
plt.ylabel('quality')
```

```python
wineDB = wineDB.drop(['pH','free sulfur dioxide','residual sugar'], axis=1)
wineDB.head()
```

```python
wineDB.isnull().sum()
```

```python
plt.figure(figsize=(14,10))
sns.countplot(wineDB['quality']);
```

```python
wineDB['quality'] = wineDB['quality'].replace(8,7)
wineDB['quality'] = wineDB['quality'].replace(3,5)
wineDB['quality'] = wineDB['quality'].replace(4,5)
wineDB['quality'].value_counts()
```

```python
x_train, x_test, y_train, y_test = train_test_split(wineDB.drop('quality',axis=1), wineDB['quality'], test_size=0.3, random_state=22)
x_train.shape, x_test.shape
```

```python
wine_cl = DecisionTreeClassifier(criterion="entropy", max_depth=3)
wine_cl = wine_cl.fit(x_train,y_train)
y_pred = wine_cl.predict(x_test)
```

```python
from sklearn import metrics 
print("Accuracy: ", metrics.accuracy_score(y_test,y_pred)*100)
```

```python
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
