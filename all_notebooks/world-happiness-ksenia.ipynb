```python
data6 = dict(type = 'choropleth', 
           locations = hap['Country'],
           locationmode = 'country names',
           z = hap['Happiness.Rank'], 
           text = hap['Country'],
          colorscale = 'Viridis', reversescale = False)
layout = dict(title = 'Happiness Rank World Map', 
             geo = dict(showframe = False, 
                       projection = {'type': 'Mercator'}))
choromap6 = go.Figure(data = [data6], layout=layout)
iplot(choromap6)
```

The map demonstrates the geograghy of happy contries.

The darker is the color - the happier is a country.

There are large differences across countries. 

According to 2017 numbers, Nordic countries top the ranking: Denmark, Iseland, the Netherlands, Switzerland and Finland  have the highest scores. North America, Australia and New Zealand are also on the top.

 The higest ranks correspond to countries of Afica and South Asia: Central African Republic, Syria, Tanzania, South Sudan, Rwanda and others..



```python
trace4 = go.Scatter(
    x = hap["Economy..GDP.per.Capita."],
    y = hap["Happiness.Rank"],
    mode = 'markers'
)
data4 = [trace4]
layout = go.Layout(
    title='Happiness Rank Determined by Economy',
    xaxis=dict(
        title='Economy, GDP',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Happiness Rank',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)

fig4 = go.Figure(data=data4, layout=layout)
iplot(fig4)
```

We can see a negative relationship between the level of economy of the country and a rank of happiness, it means the higher is GDR - the lower is the rank and happier people who lives in the country. Still there is some decrepentcy so it is not an exact linear correlation but very close.
