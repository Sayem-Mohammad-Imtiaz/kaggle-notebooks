```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('../input/suicide-rates-overview-1985-to-2016/master.csv')
base=pd.DataFrame(base)
base.head()
```

```python

base1=base.drop(['HDI for year'], axis=1)
base1.head()
```

```python
base1['suicides_no'].describe()
```

```python
sns.heatmap(base1.corr(),cmap='coolwarm',annot=True)
plt.show()
```

```python
corr2 = sns.PairGrid(base1)
corr2.map(plt.scatter)
plt.show()
```

Con los anteriores dos gráficos se pretende conocer la relación entre las variables,el primero nos permite ver de forma númerica su correlación además de representarlos por distintos colores.El segundo, se utiizó para aprecisar si existe alguna relación no lineal entre las variables. Cabe resaltar, la correlación negativa entre número de suicidios y año, también,la correlación positiva aunque débil entre número de suicidios y GDP per capita. 

```python
plt.figure(figsize=(15,5))
graf1 = sns.barplot(x = 'sex', y = 'suicides_no', hue = 'age',data = base1,  palette='gnuplot2_r')
plt.xlabel=('Género')
graf1.set(ylabel="número de suicidios",xlabel="Género")
```

En esta grafica observamos el número de suicidios por género en las distintas edades, como observamos la edad 
en la cual se presentan mayor número de suicidios para hombres es de 35-54 años, para mujeres es la misma edad.
Sin embargo,es menor la porporción de suicidios en mujeres que en hombres. 

```python
plt.figure(figsize=(15,5))
graf2 = sns.barplot(x = 'generation', y = 'suicides_no',data = base1, palette="RdYlBu_r")
plt.title('Número de suicidios por Generación')
graf2.set(ylabel="número de suicidios",xlabel="Generación")
```

Para esta gráfica, tenemos el número de suicidios por generación. Como observamos la genéración con mayor número de suicidios es la Boomemers, que es la genaración de personas nacidas entre 1945 y 1964. Y con un menor número de suicidios se encuentra la generación Z, nacidos entre 1995 hasta el presente. 

```python
plt.figure(figsize=(15,5))
graf3 = sns.lineplot(x = 'year', y = 'suicides/100k pop',data = base1)
plt.title('Suicidios/100k pop por año')
graf3.set(ylabel='Suicidios/100k',xlabel='Años')
```

En esta gráfica, tenemos el número de suicicios por cada 100.000 habitantes por año. Se observa un pico después del año 1995 lo cual se puede relacionar con la generación boomers. Además, se ve una tendencia a la alta después del año 2015. 

```python
sns.lmplot("suicides_no", "gdp_per_capita ($)", data=base1)
plt.show()
```

Para esta gráfica, se prentende conocer la relación entre la variable GDP per capita con el número de suicidios. Es interesante puesto que notamos que los países con mayor GDP per capita cuentan con un número de suicicios de casi cero, por el contrario, los paises con menos GDP per capita tienen el mayor número de suicidios. No obstante, existe una correlación positiva muy débil entre los paises con un GDP medio bajo con el número de suicidios. 

```python
sns.kdeplot(base1['suicides_no'])
plt.show()
```

Con esta gráfica, se pretendía conocer una aproximación continua de la distribución de densidad para el número de suicidios. 

```python
plt.figure(figsize=(10,20))
graf7 = sns.barplot(x = 'suicides/100k pop', y = 'country',data = base1,  palette='Accent_r')
plt.xlabel=('Género')
graf7.set(ylabel="País",xlabel='Suicidios por 100.000 pop')

```

En esta gráfica, se observa el número de suicidios por cada 100.000 habitantes en promedio por cada país. El país que presenta mayor número de suicidios por cada 100.000 habitantes en promedio es Lithuania. Se debe tener en cuenta que este promedio se realiza desde el 1985 hasta el 2016.
