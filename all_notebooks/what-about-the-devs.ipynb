```python
data = pd.read_csv('../input/2017-fCC-New-Coders-Survey-Data.csv')
```

```python
# Which are the "top" 15 countries in terms of number of developers per country?
most_devs_country = data.groupby(by='CountryLive').size().sort_values(ascending=False).head(15)
# Most common dev ages 
most_common_ages = data.groupby(by='Age').size().sort_values(ascending=False).head(15)
```

```python
countries_gender = data.groupby(by=['CountryLive', 'Gender']).size()
# Countries with most female devs
countries_gender.loc[countries_gender.index.isin(['female'], level='Gender')].sort_values(ascending=False).head(15)
```

```python
expected_earnings_gender = data.groupby(by=['Gender']).agg({'ExpectedEarning': 'mean'})

expected_earnings_gender['$ from most well paid gender mean'] = expected_earnings_gender['ExpectedEarning'] - expected_earnings_gender['ExpectedEarning'].max()
expected_earnings_gender
```
