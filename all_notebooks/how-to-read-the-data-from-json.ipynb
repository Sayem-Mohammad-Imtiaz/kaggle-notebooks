```python
import json

with open('/kaggle/input/covid19-data-of-india/covid_cases.json') as file:
    my_file=json.load(file)
    
timestamp=[]
cases=[]

for i in my_file:
    temp = my_file[i]
    i = int(i)
    timestamp.append(i)
    cases.append(int(temp))
    
print("Timestamp list:- "+str(timestamp))
print("Cases list:- "+str(cases))
```
