```python
x,y = df.drop("Species",axis=1),df["Species"]
```

```python
import sklearn.preprocessing as process
import pandas_profiling as pp
```

```python
profile = pp.ProfileReport(df)
profile.to_file("index.html")
profile 
```

```python
y = process.LabelEncoder().fit_transform(y)
```

```python
x.head()
```

```python
y[:5]
```

```python
from sklearn.linear_model import LogisticRegression
import pickle
```

```python
mod = LogisticRegression().fit(x,y)
y_pred = mod.predict(x)
count = 0
for i,j in zip(y,y_pred):
    if i != j:
       count += 1
print("error {}%".format(count/len(y)*10))
```

```python
with open("./model.pkl","wb") as f:
    pickle.dump(mod,f)
```

```python
print("the weight are {} and biases are {}".format(mod.coef_,mod.intercept_))
```
