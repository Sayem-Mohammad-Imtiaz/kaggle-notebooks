```python
#reading the data

reviews = pd.read_csv('/kaggle/input/womens-ecommerce-clothing-reviews/Womens Clothing E-Commerce Reviews.csv')
reviews.head(3)
```

```python
#Creating a dataframe with relevant columns

df = reviews[['Review Text', 'Recommended IND']]


#checking for missing values
df.isna().sum()

#removing rows with missing values for 'Review Text'
df1 = df.dropna(axis=0, how='any')
```

```python
#creating an n-gram (1&2) model using 'Recommended IND as a proxy for sentiment'

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split

vectorizer = CountVectorizer(stop_words = 'english', lowercase = True, ngram_range=(1,2))

X_train, X_test, y_train, y_test = train_test_split(df1['Review Text'], df1['Recommended IND'])

X_train_bow = vectorizer.fit_transform(X_train)
X_test_bow = vectorizer.transform(X_test)
```

```python
#Building a Naive Bayes classifier to predict sentiment

from sklearn.naive_bayes import MultinomialNB

clf = MultinomialNB()

#train the model 
clf.fit(X_train_bow, y_train)

#Measure accuracy on test set
accuracy = clf.score(X_test_bow, y_test)
print(accuracy)
 
```
