# Spotify Analysis

```python
import numpy as np
import pandas as pd
import plotly.express as px

from matplotlib import rcParams
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import mean_squared_error

rcParams['figure.figsize'] = 10,7.5
rcParams['figure.dpi'] = 80
```

Can check this notebook in: https://www.kaggle.com/joaquincastaon/assestment2-ucsp-spotify-analysis

```python
df = pd.read_csv('../input/spotify-dataset-19212020-160k-tracks/data.csv')
df
```

```python
ts_df = df.groupby("year")[['year',"acousticness","danceability","energy","speechiness","liveness","valence"]].mean()
fig = px.line(ts_df, x='year', y=ts_df.columns,title='Audio Characteristics')
fig.show()
```

El porcentaje de la caracteristica acustica decrese a partir del año 1953, porque comenzo a crecer la musica electronica en el año 51, con lo que lleva a que la caracteristica de energia en la musica comience a tener relevancia.

La caracteristica valence que representa la felicidad en las canciones se mantuvo constante con ciertas disminuciones a partir del año 2000 hasta el año 2017 se observa un disminucion, que puede deberse,a la presencia del genero emo-rock.

A partir del año 2011 se observa una tendencia a crecer de la caracteristica bailable de las canciones, debido al aumento de artistas de bachata y reggaeton y musico electro.

En cuanto a la caracteristica 'speechiness' se pouede observar una disminucion significativa en el año 1954, para luego comenzar a tomar fuerza al rededor de los años 90s debido a la aparicion del genero hiphop y rap entre otros similares.

```python
data_by_genres = pd.read_csv('../input/spotify-dataset-19212020-160k-tracks/data_by_genres.csv')
data_by_genres
```

```python
group_genre_charct = data_by_genres.groupby('genres')[["acousticness","danceability","energy","speechiness","liveness","valence", 'popularity']].mean()
group_genre_charct = pd.DataFrame(group_genre_charct).reset_index()
genre_popularity = group_genre_charct[['genres','popularity']].sort_values(by= 'popularity', ascending = False)
genre_popularity.head(20)

fig, ax= plt.subplots()
sns.barplot(x="popularity", y = "genres", data = genre_popularity.head(10), palette = 'mako')
ax.set_xlim(70, 82)

plt.show()
```

Entre los 10 generos más populares en el mundo son: basshall, turkish edm, south african house, trap venezolano entre otros.

```python
duration_year = df.groupby('year')['year','duration_ms'].mean()
duration_year['duration_ms'] = duration_year['duration_ms']/60000
fig = px.line(duration_year, x='year', y='duration_ms',title='Duration of Songs')
fig.show()
```

En el año 1946 se observa que el promedio de duracion de las canciones, tuvo un pico cuyo valor fue de 4 minutos y medio aproximadamente, sin embargo hubo un lapso de tiempo entre los años 60s y 70s donde la duracion disminuyo considerablemente llegando a valores de 3 minutos aproximadamente.

Desde el año 71 hasta el 2012 se observa que la duracion promedio de las canciones tiene una disminucion lenta que va desde los 4 minutos y medio de duracion hasta 4 minutos.

Despues del 2012 se observa una disminucion drastica hasta el año actual en el que se disminuye casi 1 minuto de duracion en las canciones.

```python
g = sns.scatterplot(x=df["duration_ms"]/60000, y="popularity",
                hue="year",
                palette="ch:r=-.2,d=.3_r",
                sizes=(20, 200), linewidth=0,
                data=df)

g.legend(loc='upper left', bbox_to_anchor=(1, 0.5))
g.set(xlabel = 'duration (min)')
```

Las canciones mas populares pertenecen al año 2000, y tienen una corta duracion de aproximadamente 3 a 4 minutos.

```python
g = sns.scatterplot(x=df["duration_ms"]/60000, y="popularity",
                hue="key", size='mode',
                palette="ch:r=-.2,d=.3_r",
                sizes=(20, 200), linewidth=0,
                data=df, alpha=0.5)

g.legend(loc='upper left', bbox_to_anchor=(1, 0.5))
g.set(xlabel = 'duration (min)')
```
