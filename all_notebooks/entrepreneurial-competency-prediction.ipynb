```python
#4. Processing feartures and target.
#IndividualProject
ind_map = {'Yes':1, 'No':2}
data['IndividualProject'] = data['IndividualProject'].map(ind_map)
```

```python
#gender
data["Gender"].unique()
```

```python
gen_map = {'Male':1, 'Female':2}
data['Gender'] = data['Gender'].map(gen_map)
```

```python
#City
cit_map = {'Yes':1, 'No':2}
data['City'] = data['City'].map(cit_map)
```

```python
#Influenced
inf_map = {'Yes':1, 'No':2}
data['Influenced'] = data['Influenced'].map(inf_map)
```

```python
#MentalDisorder
men_map = {'Yes':1, 'No':2}
data['MentalDisorder'] = data['MentalDisorder'].map(men_map)
```

```python
#KeyTraits
data["KeyTraits"].unique()
```

```python
key_map = {'Passion':1, 'Vision':2,'Resilience':3, 'Positivity':4,'Work Ethic':5}
data['KeyTraits'] = data['KeyTraits'].map(key_map)
```

```python
data.head()
```

```python
#5. Determining x and y
x=data[["IndividualProject","Age","Gender","City", "Influenced","Perseverance","DesireToTakeInitiative","Competitiveness","SelfReliance","StrongNeedToAchieve","SelfConfidence","GoodPhysicalHealth","MentalDisorder","KeyTraits"]]
x.head()
```

```python
y=data["y"]
y.head()
```

```python
#6. splitting data into train and test
x_train, x_test, y_train, y_test=train_test_split(x, y, random_state=6)
##standardizing data
transfer = StandardScaler()
x_train = transfer.fit_transform(x_train)
x_test = transfer.transform(x_test)
```

```python
# Using different machine learning methods
#Using Random Forest
estimator=RandomForestClassifier(criterion='entropy', max_depth=7, n_estimators=10)
#estimator=RandomForestClassifier(n_estimators=10, criterion="entropy", max_depth=8, bootstrap=True, max_features="auto")
estimator.fit(x_train, y_train)
y_predict=estimator.predict(x_test)
accuracy = estimator.score(x_test, y_test)
print(accuracy)
```

```python
#Using Xgboost
estimator= XGBClassifier()
estimator.fit(x_train, y_train)
y_predict=estimator.predict(x_test)
accuracy = estimator.score(x_test, y_test)
print("The accuracy by Xgboost:\n", accuracy)
```

```python
#Using KNN
estimator = KNeighborsClassifier(n_neighbors=3)
estimator.fit(x_train,y_train)
y_predict=estimator.predict(x_test)
accuracy = estimator.score(x_test, y_test)
print(accuracy)
```

```python
#Based on the accuracy, I choose random forest to adjust its params
#Using Random Forest
estimator=RandomForestClassifier(criterion='entropy', max_depth=7, n_estimators=10)
#estimator=RandomForestClassifier(n_estimators=10, criterion="entropy", max_depth=8, bootstrap=True, max_features="auto")
estimator.fit(x_train, y_train)
y_predict=estimator.predict(x_test)
accuracy = estimator.score(x_test, y_test)
print(accuracy)
param_dict = {'n_estimators': [10, 20, 30, 40, 50, 60, 70], 'max_depth':range(3,14,2)}
estimator = GridSearchCV(estimator, param_grid=param_dict, cv=10)
estimator.fit(x_train,y_train)
print("最佳参数:\n", estimator.best_params_)
print("最佳估计器：\n", estimator.best_estimator_)
print("最佳分数:\n", estimator.best_score_)
```

```python
estimator=RandomForestClassifier(criterion='entropy', max_depth=3, n_estimators=10)
#estimator=RandomForestClassifier(n_estimators=10, criterion="entropy", max_depth=8, bootstrap=True, max_features="auto")
estimator.fit(x_train, y_train)
y_predict=estimator.predict(x_test)
accuracy = estimator.score(x_test, y_test)
print(accuracy)
param_dict = {'min_samples_split':range(50,201,20), 'max_features':range(3,11,2)}
estimator = GridSearchCV(estimator, param_grid=param_dict, cv=10)
estimator.fit(x_train,y_train)
print("最佳参数:\n", estimator.best_params_)
print("最佳估计器：\n", estimator.best_estimator_)
print("最佳分数:\n", estimator.best_score_)
```

If you want to be an entrepreneur, you can try to be close to the factors listed in the research. However, gainning all the factors promoting the entreprenurial competency can not guarantee that you can become a entrepreneur.
