```python
test=pd.read_csv("../input/ec-rawdata/test.csv")
train=pd.read_csv("../input/ec-rawdata/train.csv")
val=pd.read_csv("../input/ec-rawdata/val.csv",)
```

```python
print(train.Rating.unique())
print(val.Rating.unique())
print(test.Rating.unique())
train.head()

column_contain1=['Title']
column_contain2=['Positive Feedback Count']
column_contain3=['Review Text']
#train set
x1_train=pd.DataFrame(data=train['Title'],columns=column_contain1)
x2_train=pd.DataFrame(data=train['Review Text'],columns=column_contain3)
y_train=pd.DataFrame(data=train['Positive Feedback Count'],columns=column_contain2)
#validation set
x1_val=pd.DataFrame(data=val['Title'],columns=column_contain1)
x2_val=pd.DataFrame(data=val['Review Text'],columns=column_contain3)
y_val=pd.DataFrame(data=val['Positive Feedback Count'],columns=column_contain2)
#test set
x1_test=pd.DataFrame(data=test['Title'],columns=column_contain1)
x2_test=pd.DataFrame(data=test['Review Text'],columns=column_contain3)
y_test=pd.DataFrame(data=test['Positive Feedback Count'],columns=column_contain2)
```

```python
y_train.describe()    #Positive Feedback Count應分類較好
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize']=(15,6)
pd.value_counts(y_train['Positive Feedback Count']).plot.bar(width=1,align='center', alpha=1)
plt.title('Original Count histogram')
plt.xlabel('Positive Feedback Count')
plt.ylabel('Frequency')
y_train['Positive Feedback Count'].value_counts()


```

```python
#分類Positive Feedback Count

y_train = y_train['Positive Feedback Count']
for i in y_train.index:
    if (y_train[i]>=1) and (y_train[i]<=3):
        y_train[i]=1
    if y_train[i]>3:
        y_train[i]=2

y_val = y_val['Positive Feedback Count']
for i in y_val.index:
    if (y_val[i]>=1) and (y_val[i]<=3):
        y_val[i]=1
    if y_val[i]>3:
        y_val[i]=2

y_test = y_test['Positive Feedback Count']
for i in y_test.index:
    if (y_test[i]>=1) and (y_test[i]<=3):
        y_test[i]=1
    if y_test[i]>3:
        y_test[i]=2   
```

```python
#文字轉向量
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(
max_features=300,
stop_words='english',
ngram_range=(1, 2))

vectorizer2 = CountVectorizer(
max_features=500,
stop_words='english',
ngram_range=(1, 2))

x1_train = x1_train['Title']
x2_train = x2_train['Review Text']
x1_val = x1_val['Title']
x2_val = x2_val['Review Text']
x1_test = x1_test['Title']
x2_test = x2_test['Review Text']


x1_train = vectorizer.fit_transform(x1_train.values.astype('U'))
x2_train = vectorizer2.fit_transform(x2_train.values.astype('U'))
x1_val = vectorizer.fit_transform(x1_val.values.astype('U'))
x2_val = vectorizer2.fit_transform(x2_val.values.astype('U'))
x1_test = vectorizer.fit_transform(x1_test.values.astype('U'))
x2_test = vectorizer2.fit_transform(x2_test.values.astype('U'))
from sklearn.feature_extraction.text import TfidfTransformer
tfidf = TfidfTransformer()
x1_train = tfidf.fit_transform(x1_train).toarray()
x2_train = tfidf.fit_transform(x2_train).toarray()
x1_val = tfidf.fit_transform(x1_val).toarray()
x2_val = tfidf.fit_transform(x2_val).toarray()
x1_test = tfidf.fit_transform(x1_test).toarray()
x2_test = tfidf.fit_transform(x2_test).toarray()
```

```python
#SMOTE解決imbalanced dataset

from imblearn.over_sampling import SMOTE
print('Shape of x1_train: {}'.format(x1_train.shape))
print('Shape of y_train: {}'.format(y_train.shape))

print("Before OverSampling, counts of label '2': {}".format(sum(y_train==2)))
print("Before OverSampling, counts of label '1': {}".format(sum(y_train==1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train==0)))

sm = SMOTE(random_state=2)
x1_train_res, y_train_res = sm.fit_sample(x1_train, y_train.ravel())


print('After OverSampling, the shape of x1_train: {}'.format(x1_train_res.shape))
print('After OverSampling, the shape of y_train: {} \n'.format(y_train_res.shape))

print("After OverSampling, counts of label '2': {}".format(sum(y_train_res==2)))
print("After OverSampling, counts of label '1': {}".format(sum(y_train_res==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res==0)))
```

```python
from imblearn.over_sampling import SMOTE
print('Shape of x2_train: {}'.format(x2_train.shape))
print('Shape of y_train: {}'.format(y_train.shape))

print("Before OverSampling, counts of label '2': {}".format(sum(y_train==2)))
print("Before OverSampling, counts of label '1': {}".format(sum(y_train==1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train==0)))

sm = SMOTE(random_state=2)
x2_train_res, y_train_res = sm.fit_sample(x2_train, y_train.ravel())


print('After OverSampling, the shape of x2_train: {}'.format(x2_train_res.shape))
print('After OverSampling, the shape of y_train: {} \n'.format(y_train_res.shape))

print("After OverSampling, counts of label '2': {}".format(sum(y_train_res==2)))
print("After OverSampling, counts of label '1': {}".format(sum(y_train_res==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res==0)))
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
clf = tree.DecisionTreeClassifier()
ec1_clf = clf.fit(x1_train,y_train)

val_x1_predicted = ec1_clf.predict(x1_val)

print('Algorithm:tree')
print('由Title預測\n')
print('*Before SMOTE*\n')
print(accuracy_score(y_val,val_x1_predicted ))
print(confusion_matrix(y_val,val_x1_predicted))
print('\n')
print(classification_report(y_val,val_x1_predicted))
```

```python
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
clf = tree.DecisionTreeClassifier()
ec1_clf = clf.fit(x1_train_res,y_train_res)

val_x1_predicted = ec1_clf.predict(x1_val)

print('Algorithm:tree')
print('由Title預測\n')
print('*After SMOTE*\n')
print(accuracy_score(y_val,val_x1_predicted ))
print(confusion_matrix(y_val,val_x1_predicted))
print('\n')
print(classification_report(y_val,val_x1_predicted))
```
