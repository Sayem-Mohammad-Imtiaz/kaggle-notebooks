**In this notebook, I am going to build a multivariate linear regression using Ordinary least square method. To build this model, I will use a dataset of 73 students. It includes attendance, Mid semester exam marks, number of hours he/she studied one day before the exam, and the marks of the End semester exam.**

First of all, I will import some usefull libraries of python. Then, I will read a csv file which contains the data of the students.

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

df = pd.read_csv('../input/students-data-for-mlr/amsPredictionSheet1-201009-150447.csv')
df
```

The following figure shows the equation of the multivariate linear regression and the simple linear ragression. We can see the basic difference from that.<br><br>
![](https://miro.medium.com/max/3444/1*uLHXR8LKGDucpwUYHx3VaQ.png)

The formula to find the parameters of the above equation is as follows. This formula is in the form of matrix and it return a matrix of all parameters. The formula ia as follows:<br>
![](https://cdn-images-1.medium.com/max/498/1*Bt1LjOBnr9vkwPTotwM2yA.png)

Now lets implement this formula in our module and get the values of all parameters. Before that, I will put all input variables and output variables in seperate matrices.

```python
y = df['ESE']
x = sm.add_constant(df[['Attendance', 'MSE', 'HRS']])
z = df[['Attendance', 'MSE', 'HRS']]
print(x, y, z)
```

```python
x_transpose = x.T
x_transpose
```

```python
a = x_transpose.dot(x)
a_inverse = np.linalg.inv(a)
a_inverse
```

```python
B = a_inverse.dot(x_transpose.dot(y))
B
```

Yes, I have got the values of parameters. So, the equation of the multivariate linear regression is as follows:

```python
def predict_y(x1, x2, x3):
    y_hat = B[1]*x1 + B[2]*x2 + B[3]*x3 + B[0]
    return y_hat;
```

The above function will predict the marks of End semester examination. Lets predict the marks of the ESE with given data.

```python
ESE_pre = predict_y(df['Attendance'], df['MSE'], df['HRS'])
ESE_pre
```

To check, how efficient our model is? Let's calculate the value of R<sup>2</sup>.

```python
r_square = 1 - sum((y-ESE_pre)*(y-ESE_pre))/sum((y-y.mean())*(y-y.mean()))
r_square
```

So, The value of R<sup>2</sup> is 0.58. So, we can say that, our model is quite efficient.
