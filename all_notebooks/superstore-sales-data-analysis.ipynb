# EXPLORATORY DATA ANALYSIS

```python
#Sales Trends
```

```python
#converting order date to timestamp
data['order_date_time'] = pd.to_datetime(data['order_date'])
```

```python
print("Minimum Date:", data['order_date_time'].min())
print("Maximum Date:", data['order_date_time'].max())
```

```python
#Extracting Month and Year from the Order_Date
data['Month_Year']=data['order_date_time'].apply(lambda x: x.strftime('%Y-%m'))
data['Month_Year'].head(5)
```

```python
#Converting Sales to float type
data['sales']=data['sales'].str.replace(',','')
data['sales']=data['sales'].astype(float)
```

```python
data.info()
```

```python
data_trend=data.groupby(['Month_Year']).sum()['sales'].reset_index()
```

```python
plt.figure(figsize=(15,6))
plt.plot(data_trend['Month_Year'],data_trend['sales'],color='#b80045')
plt.xticks(rotation='vertical',size=8)
plt.show()
```

```python
top_products=pd.DataFrame(data.groupby(['product_name']).sum()['sales'])
top_products.sort_values('sales',ascending=False).head(10)
```
