```python
num_cols=list(X.select_dtypes(['float64','int64']).columns)
```

```python
num_cols
```

```python
X.info()
```

```python
#For Imputation, Scaling
pipeline_num = Pipeline(steps=[
    ('imputer', Imputer(strategy='median')),
    ('scaling',StandardScaler())
])
```

```python
# For Encoding
pipeline_cat = Pipeline(steps=[
    ('imputer', Imputer(strategy='')),
    ('encoding', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', pipeline_num, num_cols),
        ('cat', pipeline_cat, cat_cols)])
```

```python
# Apply preprocess 
pipeline = Pipeline(steps=[('preprocessor',preprocessor),
                ('classifier',RandomForestClassifier(n_estimators=10))])
```

```python
X.shape
```

```python
train_X, test_X, train_y, test_y = train_test_split(X,y,random_state = 0)
```

```python
pipeline.fit(train_X,train_y)
```

```python
    param_grid = {
    'preprocessor__num__imputer__strategy': ['mean', 'median'],
    'classifier__n_estimators': [10,15,20]
    
}
```

```python
gs = GridSearchCV(pipeline, param_grid=param_grid, cv=5, n_jobs=-1)
```

```python
gs.fit(train_X,train_y)
```

```python
y_pred=gs.predict(test_X)
```

```python
print("Model Accuracy: ")
print(format(gs.score(test_X,test_y)))
```
