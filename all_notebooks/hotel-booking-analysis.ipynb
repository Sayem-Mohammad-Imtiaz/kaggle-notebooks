```python
df["country"].value_counts()[0:7] #which countries do have the most number of travelers?Identify the five countries having the most number of travelers
```

```python
new=df[df["total_of_special_requests"]==5]
new[["name","email","total_of_special_requests"]]
```

```python
#The most frequent last name:
def family(name:str):  
    return name.split()[1]
df["family"]= df["name"].apply(family)
df["family"].value_counts()[:5]
```

```python
#People whith the most children and babies:
df["kids"]= df["children"]+df["babies"]
df.groupby("name").sum()["kids"].sort_values(ascending=False).head()
```

```python
# The most number of local calling codes:
df["local_code"] = df["phone-number"].apply(lambda code:code[:4])
df["local_code"].value_counts()[:4].reset_index()
```

> **Matrix Heatmap On Year | Month**

```python
c = df.groupby("arrival_date_year")["arrival_date_month"].value_counts().unstack().fillna(0)
sns.heatmap(data=c,cmap="YlGnBu")
plt.show()
```
