```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data['status'].hist(figsize=(4, 4))
plt.xticks(rotation = 90)
plt.show()
```

```python
print(data.columns)
categorical = [i for i in data.columns if data.dtypes[i] == 'object']
print(categorical)
numericals = [i for i in data.columns if data.dtypes[i] != 'object']
numericals.pop(0)
print(numericals)
```

```python
fig, ax = plt.subplots(figsize=(8,8))
sns.boxplot(x = 'specialisation', y = 'salary', data = data, showfliers = False, hue = 'degree_t', ax = ax)
plt.show()
```

* Ignoring others stats here as it has many categories included in it
* Sci & tech degree students have higher median packages than comm & manage. students in both specialisations
* They also have higher 25th and 75th quantiles than those students

```python
data.groupby(['status'])['degree_t'].value_counts()
```

```python
_, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

sns.countplot(x = 'status', data = data, hue = 'specialisation', ax = axes[0])
sns.countplot(x = 'specialisation', data = data, hue = 'degree_t', ax = axes[1])
plt.show()
```

```python
corr_matrix = data[numericals].corr()
sns.heatmap(corr_matrix)
```
