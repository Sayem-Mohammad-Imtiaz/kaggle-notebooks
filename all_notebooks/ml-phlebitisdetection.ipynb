```python
#Separate data into input sets and output sets
X = df.drop(columns = ['VipGroup'])
#X
y = df['VipGroup']
y
```

```python
#Split data into 2 sets: training & testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
#Build a model using ML algorithm. Decision Tree
Model = DecisionTreeClassifier()
```

```python
#Train the model so that it learns pattern in data
#Model.fit(X, y)
Model.fit(X_train, y_train)
```

```python
#Visualizing Decision Tree
tree.export_graphviz(Model, out_file = 'VipGroup-predictor.dot',
                    feature_names = ['VipScore', 'TempInsertionSite'],
                    class_names = sorted(y.unique()),
                    label = 'all',
                    rounded = True,
                    filled = True)
```
