```python
import pandas as pd
import numpy as np
import sklearn
import matplotlib.pyplot as plt
import math
```

```python
d = pd.read_csv("../input/flight-take-off-data-jfk-airport/M1_final.csv")
d
```

```python
d.info()
```

**LABEL ENCODING**

```python
from sklearn.preprocessing import LabelEncoder 
label = LabelEncoder()

d['OP_UNIQUE_CARRIER'] = label.fit_transform(d['OP_UNIQUE_CARRIER'])
d['TAIL_NUM'] = label.fit_transform(d['TAIL_NUM'])
d['DEST'] = label.fit_transform(d['DEST'])
d['Dew Point'] = label.fit_transform(d['Dew Point'])
d['Wind'] = label.fit_transform(d['Wind'])
d['Condition']= label.fit_transform(d['Condition'])
d.head()
```

```python
Y = d["TAXI_OUT"].values
X = d.drop("TAXI_OUT",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.10,shuffle= True)
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

**LINEAR REGRESSION(LABEL ENCODING)**

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

le_linear = LinearRegression()
le_linear = le_linear.fit(X_train,Y_train)

le_linear_rms = (mean_squared_error(Y_test,le_linear.predict(X_test))**0.5)
print(le_linear_rms)
```

**RIDGE REGRESSION(LABEL ENCODING)**

```python
from sklearn.linear_model import Ridge

le_ridge = Ridge(alpha=0.05, normalize=True)
le_ridge.fit(X_train,Y_train)

le_ridge_rms = mean_squared_error(Y_test,le_ridge.predict(X_test))**0.5
print(le_ridge_rms)
```

**LASSO REGRESSION(LABEL ENCODING)**

```python
from sklearn.linear_model import Lasso

le_lasso = Lasso(alpha=0.05, normalize=True)
le_lasso.fit(X_train,Y_train)

le_lasso_rms = mean_squared_error(Y_test,le_lasso.predict(X_test))**0.5
print(le_lasso_rms)
```

**KNN MODEL(LABEL ENCODING)**

```python
from sklearn.neighbors import KNeighborsRegressor

le_KNN = KNeighborsRegressor(n_neighbors = 200)
le_KNN.fit(X_train, Y_train)

le_KNN_rms = mean_squared_error(Y_test,le_KNN.predict(X_test))**0.5
print(le_KNN_rms)
```

**SVR(LABEL ENCODING)**

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler

sc1 = StandardScaler()
sc2 = StandardScaler()

x1 = X_train.copy()
x2 = X_test.copy()

x1 = sc1.fit_transform(x1)
x2 = sc2.fit_transform(x2)

svr = SVR()
svr.fit(x1,Y_train)

le_svr_rms = mean_squared_error(Y_test,svr.predict(x2),squared = False)
print(le_svr_rms)
```

**NAIVE BAYES(LABEL ENCODING)**

```python
from sklearn.linear_model import BayesianRidge

le_bayes = BayesianRidge()
le_bayes.fit(X_train, Y_train)

le_bayes_rms = mean_squared_error(Y_test,le_bayes.predict(X_test))**0.5
print(le_bayes_rms)
```

**RANDOM FOREST(LABEL ENCODING)**

```python
from sklearn.ensemble import RandomForestRegressor

le_random = RandomForestRegressor()
le_random.fit(X_train,Y_train)

le_random_rms = mean_squared_error(Y_test,le_random.predict(X_test))**0.5
print(le_random_rms)
```

**LIGHT GBM(LABEL ENCODING)**

```python
from lightgbm import LGBMRegressor

le_gbm = LGBMRegressor()
le_gbm.fit(X_train,Y_train)

le_gbm_rms = mean_squared_error(Y_test,le_gbm.predict(X_test))**0.5
print(le_gbm_rms)
```

**ONE HOT ENCODING**

```python
print(d['OP_UNIQUE_CARRIER'].unique())
print(d['TAIL_NUM'].unique()) 
print(d['DEST'].unique())
print(d['Dew Point'].unique())
print(d['Wind'].unique())
print(d['Condition'].unique())
d['DEST'].value_counts()
```
