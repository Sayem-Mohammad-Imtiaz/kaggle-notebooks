```python
import pandas as pd
import plotly.express as px
import statistics
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
Data = pd.read_csv('../input/list-of-700-unicorn-companies-across-the-globe/List of 700 Unicorn Companies.csv')
Data1 = Data
```

```python
Data.info()
```

```python
Data["Date Joined"] = pd.to_datetime(Data["Date Joined"])
Data['Year'] = pd.DatetimeIndex(Data['Date Joined']).year
```

```python
for column in Data:
    print("\n" + column)
    print(len(Data[column].unique()))
```

```python
Data["Valuation (in $B)"]=Data["Valuation (in $B)"].str.replace('$','')
```

```python
Data["Valuation (in $B)"]=Data["Valuation (in $B)"].astype(float)
```

```python
Data['Country of origin'] = Data['Country of origin'].replace({'Austria':'other' })
Data['Country of origin'] = Data['Country of origin'].replace({'Philippines':'other'})
Data['Country of origin'] = Data['Country of origin'].replace({'Croatia':'other'})
Data['Country of origin'] = Data['Country of origin'].replace({'Thailand':'other'})
Data['Country of origin'] = Data['Country of origin'].replace({'Lithuania':'other'})
Data['Country of origin'] = Data['Country of origin'].replace({'Luxembourg':'other'})
Data['Country of origin'] = Data['Country of origin'].replace({'Belgium':'other'})
Data['Country of origin'] = Data['Country of origin'].replace({'Finland':'other'})
Data['Country of origin'] = Data['Country of origin'].replace({'Turkey':'other'})
Data['Country of origin'] = Data['Country of origin'].replace({'Estonia':'other'})
Data['Country of origin'] = Data['Country of origin'].replace({'Norway':'other'})
Data['Country of origin'] = Data['Country of origin'].replace({'Bermuda':'other'})
```

```python
Data['Industry'] = Data['Industry'].replace({'Healh':'Health'})
Data['Industry'] = Data['Industry'].replace({'Artificial Intelligence':'Artificial intelligence'})
Data['Industry'] = Data['Industry'].replace({'Data management & analytics<':'Data management & analytics'})
Data['Industry'] = Data['Industry'].replace({'Edtech':'Education'})
```

```python
Data["Industry"].value_counts()
```

```python
#NullValues with the help of Heatmap
sns.heatmap(Data.isnull(), cmap="inferno")
plt.show()
```

```python
null = Data.isnull().sum().sort_values(ascending=False)
#First sum and order all null values for each variable
percent = (Data.isnull().sum()/Data.isnull().count()).sort_values(ascending=False) 
dataType = Data.dtypes
missing_variables = pd.concat([null, percent,dataType], axis=1, keys=['Total', 'percent','Data Type'])
missing_variables = missing_variables.head(19)
```

```python
group = Data.groupby(["Industry"])["Valuation (in $B)"]

group.agg(['sum','mean','median','max','count']).reset_index(drop=False)
```

```python
Industry = pd.DataFrame(group.mean().reset_index(drop=False))
```

```python
Industry.plot(x="Industry",y="Valuation (in $B)",kind = 'barh',title='Average Industry Valuation')
plt.show()
```

### On an Average Auto & Transportation is rated best among 700 unicorn Projects

```python
IndustryT = pd.DataFrame(group.sum().reset_index(drop=False))
```

```python
IndustryT.plot(x="Industry",y="Valuation (in $B)",kind = 'barh',title='Total Industry Valuation')
plt.show()
```

### Fintech is rated best in terms of total market valuation 

```python
grouped = Data.groupby("Country of origin")["Valuation (in $B)"]

grouped.agg(['sum','mean','median','max','count']).reset_index(drop=False)
```
