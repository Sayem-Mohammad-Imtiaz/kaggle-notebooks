# Data Analysis & Visualization
As our data is ready, we'll do our analysis

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mpl_dates
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.bar(df['year'], df['islmsun'], color='Red')
ax.set_xlabel('Year', fontsize='11')
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
ax.set_ylabel('Sunni', fontsize='11')
ax.set_yscale('log')
plt.title('Sunni', fontsize='20')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.bar(df['year'], df['islmshi'], color='lime')
ax.set_xlabel('Year', fontsize='11')
ax.set_ylabel('Shia', fontsize='11')
ax.set_yscale('log')
plt.title('Shia', fontsize='20')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20,8))
ax.bar(df['year'], df['islmahm'])
ax.set_xlabel('Year', fontsize='11')
ax.set_ylabel('Ahmadi', fontsize='11')
ax.set_yscale('log')
plt.title('Ahmadiyya', fontsize='20')
plt.show()
```
