```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%matplotlib inline
```

```python
data=pd.read_csv("../input/Iris.csv")

print(data.head)

data.head()
```

```python
# Standardising the data

features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

x = data.loc[:, features].values
```

```python
y = data.loc[:,['Species']].values
```

```python
x = StandardScaler().fit_transform(x)
```

```python
pd.DataFrame(data = x, columns = features).head()
```

```python
# PCA projection to 2 components

pca = PCA(n_components=2)
```

```python
principalComponents = pca.fit_transform(x)

principalDf = pd.DataFrame(data = principalComponents

             , columns = ['principal component 1', 'principal component 2'])

principalDf.head(5)
```

```python
data[['Species']].head()
```

```python
finalDf = pd.concat([principalDf, data[['Species']]], axis = 1)

finalDf.head(5)
```

```python
# 2D Visualization

fig = plt.figure(figsize = (8,8))

ax = fig.add_subplot(1,1,1) 

ax.set_xlabel('Principal Component 1', fontsize = 15)

ax.set_ylabel('Principal Component 2', fontsize = 15)

ax.set_title('2 Component PCA', fontsize = 20)

targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

colors = ['r', 'g', 'b']

for Species, color in zip(targets,colors):

    indicesToKeep = finalDf['Species'] == Species

    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']

               , finalDf.loc[indicesToKeep, 'principal component 2']

               , c = color

               , s = 50)

ax.legend(targets)

ax.grid()
```

```python
# How much information (variance) in each PC

pca.explained_variance_ratio_
```
