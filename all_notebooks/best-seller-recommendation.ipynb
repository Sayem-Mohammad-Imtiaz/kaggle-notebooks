# 정확도 측정

```python
ratings = pd.read_csv('../input/movielens/ratings.csv')
movies = pd.read_csv('../input/movielens/movies.csv')

ratings = ratings.drop('timestamp', axis=1)
```

```python
from sklearn.model_selection import train_test_split
x = ratings.copy()
y = ratings['userId']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, stratify=y)

mean_ratings = x_train.groupby(['movieId'])['rating'].mean()
```

```python
def RMSE(y_true, y_pred):
    return np.sqrt(np.mean((np.array(y_true) - np.array(y_pred))**2))

def score(model):
    id_pairs = zip(x_test['userId'], x_test['movieId'])
    y_pred = np.array([model(user, movie) for (user, movie) in id_pairs])
    y_true = np.array(x_test['rating'])
    return RMSE(y_true, y_pred)
```

```python
def best_seller(user_id, movie_id):
    if movie_id in mean_ratings:
        return mean_ratings[movie_id]
    else:
        return 3.0
```

```python
score(best_seller)
```
