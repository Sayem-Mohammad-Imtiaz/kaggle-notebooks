```python
import matplotlib.pyplot as plt
import seaborn as sns


import warnings
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('/kaggle/input/rainfall-data-from-1901-to-2017-for-india/Rainfall_Data_LL.csv')
df
```

```python
df.drop(columns = ['Name'],inplace = True)
```

```python
df.SUBDIVISION.unique()
```

```python
df.info()
```

```python
#we will use subdivision, year, annual, jan-feb, mar-may, june-sept, oct-Dec columns alone 


df = df[['SUBDIVISION','YEAR','ANNUAL','Jan-Feb' , 'Mar-May', 'June-September' ,'Oct-Dec' ]]
df
```

```python
df.describe()
```

```python

conditions = [
    (df['ANNUAL'] <= 800),
    (df['ANNUAL'] > 800) & (df['ANNUAL'] <= 2000),
    (df['ANNUAL'] > 2000) & (df['ANNUAL'] <= 4500),
    (df['ANNUAL'] > 4500)
    ]


values = ['low', 'medium', 'high', 'very_high']

df['status'] = np.select(conditions, values)
df
```

```python
#scatter plot

plt.figure(figsize = (15,5))
sns.scatterplot(x ='YEAR' , y='ANNUAL', data = df)

plt.show()
```

```python
#scatter plot wit hue

plt.figure(figsize = (15,5))
sns.scatterplot(x ='YEAR' , y='ANNUAL', data = df ,hue = 'status' )

plt.show()
```

```python
sns.set_style("whitegrid")

sns.catplot(x='status', y ='SUBDIVISION', data =df , height = 10)
plt.show()
```

```python
plt.figure(figsize = (15,10))
chart = sns.boxplot(x ='SUBDIVISION' , y ='Jan-Feb' , data = df)
chart.set_xticklabels(chart.get_xticklabels(), rotation = 90)
plt.show()
```

We will try to group by year and see how the rainfall is all over INDIA from 1901 to 2017

```python
grouped_df = df.groupby('YEAR').sum()
grouped_df
```

```python
x = grouped_df[grouped_df['ANNUAL']>50000]
plt.figure(figsize = (15,10))
chart2 = sns.barplot(x = x.index , y = 'ANNUAL' , data =x)
chart2.set_xticklabels(chart2.get_xticklabels(), rotation = 90)
plt.show()
```

```python
sns.regplot(x=grouped_df.index, y="ANNUAL", data=grouped_df);
```

```python
sns.scatterplot(x = grouped_df.index , y = 'June-September' , data = grouped_df)
```

```python
sns.regplot(x = grouped_df.index , y = 'June-September' , data = grouped_df)
```

```python
sns.displot(grouped_df , x = 'YEAR' , stat = 'probability')
```

```python
sns.displot(grouped_df['ANNUAL'] , x = 'YEAR' ,kind = 'kde', bw_adjust = 0.1)
```

```python
chart3 = sns.catplot(
    data=df[df['YEAR'].isin([1910,1940,1970,2000,2015])],
    x='YEAR',
    kind ='count',
    palette='Set1',
    row='status',
    aspect=3,
    height=3
)
```
