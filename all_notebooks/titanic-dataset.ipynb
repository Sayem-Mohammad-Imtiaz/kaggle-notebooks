```python
data=data.drop(["PassengerId","Name","Ticket","Cabin", "Embarked"],axis=1)
```

Encode data 'sex' to 0 dan 1

```python
data.Sex=pd.get_dummies(data.Sex)
data.head()
```

```python
data2 = data

data2.head()
```

> We will be using int data to make easier ML process

```python
data2.info()
```

data normalization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

Drop 'Survived','Age' and 'Fare' column

```python
X=data2.drop(['Survived', 'Age','Fare'], axis=1)
```

Take 'Survived' columns set as label

```python
y=data2["Survived"]
```

```python
data2.head()
data_r=data.dropna(inplace=True)
```

fitting scaler to data

```python
X=scaler.fit_transform(X)
```

Import train_test_split for splitting dataset to data training and data testing with 70:30 ratio

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

Import Machine Learning Algoritrhm Model/Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

Fitting model to train data

```python
model.fit(X_train,y_train)
```

Predicting data test using model Logistic Regression

```python
prediction=model.predict(X_test)
```

Getting the accuracy score from the model

```python
from sklearn import metrics
```

```python
print("the accuracy of the model is : ",metrics.accuracy_score(prediction,y_test)*100)
```
