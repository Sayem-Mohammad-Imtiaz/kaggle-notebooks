A Random Forest Regression model for predicting salary of a employee according to their level, more level means higher the position of the employee and hence higher the salary.
Dataset consists of three features,first feature will be ignored because position or level is one of the same thing, hence only level feature will be taken (level is independent and salary is dependent)

```python
#importing libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing dataset
ds = pd.read_csv('../input/position-salary-dataset/Position_Salaries.csv')
ds.head()
```

```python
#since position column and level column are one of the same thing we'll take only level column as our independent variable
x = ds.iloc[:,1:-1].values
y = ds.iloc[:,-1].values
print(x)
print(y)
```

**SPLITTING THE DATASET**

* as only 10 rows are present in the dataset, no splitting will be done so that we can have more data to train our model
* splitting will decrease accuracy

```python
#training the model
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators = 10)
rf.fit(x,y)
```

```python
#visualizing predicted set result
x_new = np.arange(min(x),max(x),0.01).reshape(-1,1)
plt.scatter(x,y,c = 'red')
plt.plot(x_new,rf.predict(x_new))
plt.title('Random Forest Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
```

```python
#checking accuracy
from sklearn.metrics import r2_score
r2_score(y,rf.predict(x))
```

ACCURACY CAME 99.2% REASON FOR SUCH A HIGH ACCURACY IS THAT RANDOM FOREST IS PRONE TO OVERFITTING AND ISN'T SUITABLE FOR SMALL DATASETS.
