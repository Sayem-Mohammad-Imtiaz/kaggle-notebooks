In this task, you are given historical data: a set of 10 attributes for loan applications. For each application, it is known whether the loan was defaulted (goal = 1), or the loan was paid on time (goal = 0).

Description of features in the X_train and X_test files:

age - borrower's age
month_income - borrower's monthly income
credit_card_utilization - the ratio of the account balance to the credit limit
duty_to_income - debt to income ratio
credits_loans - current number of loans and credit lines
mortgage_loans - current number of mortgage loans
family_members - the number of members in the borrower's family
overdue_30_59_days - the number of previous overdue loan payments from 30 to 59 days
overdue_60_89_days - the number of previous overdue loan payments from 60 to 89 days
overdue_more_than_90_days - number of previous overdue loan payments for more than 90 days
purpose - whether the loan payments were overdue for more than 90 days - only available in the training sample
The y_train file contains the answers whether or not there was a defaulted (0 or 1) in fact
