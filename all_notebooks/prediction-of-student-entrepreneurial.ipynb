MODEL CREATION

```python
df = pd.get_dummies(df)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
X = df.drop(['y'], axis = 1)
y = df['y']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

```python
print(f"Training data shape: {X_train.shape}, Training data label shape: {y_train.shape}")
print(f"Testing data shape: {X_test.shape}, Testing data label shape: {y_test.shape}")
```

```python
ss = StandardScaler()
X_train_s = ss.fit_transform(X_train)
X_test_s = ss.fit_transform(X_test)
```

```python
lda = LinearDiscriminantAnalysis()
lda.fit(X_train_s, y_train)
pred = lda.predict(X_test_s)
```

```python
print(f"Accuracy score of LinearDiscriminant model: {accuracy_score(y_test, pred)}")
```

```python
from sklearn.svm import SVR

svr = SVR()
svr.fit(X_train_s, y_train)
svr_pred = svr.predict(X_test_s)
```

```python
#mean_absolute_error(y_test, svr_pred)
print(f"Mean Squared error of the given model: {mean_squared_error(y_test, svr_pred)}")
```
