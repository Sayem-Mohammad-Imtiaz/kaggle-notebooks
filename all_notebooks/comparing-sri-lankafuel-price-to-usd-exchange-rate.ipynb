```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Reading fuel prices and exchange rates

```python
exchange_rates = pd.read_csv('../input/usd-lkr-weekly-exchange-rates19902021/USD_LKR Historical Data.csv', parse_dates=True, index_col="Date")
exchange_rates.head(5)
```

```python
fuel_prices = pd.read_csv('../input/sri-lanka-fuel-prices-1992-to-2020/sl_petrol_prices.csv', parse_dates=True, index_col="Date").iloc[:, 1:]
fuel_prices.head(5)
```

```python
plt.figure(figsize=(18,6))
sns.lineplot(x ='Date' , y='LP92' , data = fuel_prices , label='Petrol price')
sns.lineplot(x ='Date' , y='Price' , data = exchange_rates , label='USD in LKR')
plt.legend()
```
