```python
# DECISION TREE CLASSIFIER ON 'heart_disease' DATASET TO COMPARE ORIGINAL AND PREDICTED VALUE.
# ALSO CALCULATING THE ACCURACY RATE.
# ACCURACY SCORE IS 100% IF RELATABLE COLUMN IS FROM 'age' TO 'thal'

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import model_selection

df=pd.read_csv('../input/heart-disease-patients-details/heart_disease.csv')

Y=df['result']
X=df.loc[:,'age':'thal']

X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,random_state=7,test_size=0.33)

dt=DecisionTreeClassifier(criterion='entropy',random_state=0)
dt.fit(X_train,Y_train)

print(dt.score(X_train,Y_train))

prediction=dt.predict(X_test)
print(prediction)
print(Y_test)
```

```python
# DECISION TREE CLASSIFIER USING STANDARD SCALER IS USED TO OBTAIN MORE ACCURATE DATA.
# DECISION TREE CLASSIFIER ON 'heart_disease' DATASET TO COMPARE ORIGINAL AND PREDICTED VALUE.
# ALSO CALCULATING THE ACCURACY RATE.
# ACCURACY SCORE IS 100% IF RELATABLE COLUMN IS FROM 'age' TO 'thal'

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import model_selection
from sklearn.preprocessing import StandardScaler 

df=pd.read_csv('../input/heart-disease-patients-details/heart_disease.csv')

Y=df['result']
X=df.loc[:,'age':'thal']

X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,random_state=7,test_size=0.33)

sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)

dt=DecisionTreeClassifier(criterion='entropy',random_state=0)
dt.fit(X_train,Y_train)

print(dt.score(X_train,Y_train))

prediction=dt.predict(X_test)
print(prediction)
print(Y_test)
```
