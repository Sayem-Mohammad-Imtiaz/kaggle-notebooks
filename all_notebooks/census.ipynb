**Feature Engineering**

Feature engineering is one of the key aspect of data science, in order to explore more data if we need to create another dimension in the dataset, we call it as feature engineering. Here for this data we will be doing the feeature engineering. Population density is one of the key role in order to study the demographics. Population density gives us the insights about how the population is distributed in all the states. Our dataset doesnt have the areas of different states, hence we will be creating another dataframe for the states and their areas. (Source: wikipedia.com) we will be importing the newly generated dataset from the csv file Area.csv. This file will be used to get the areas of states which we will be using as the denominator for total persons for that particular state. In the end we will be plotting density(PDF) function.

```python
Area=pd.read_csv('../input/area-state/State_area.csv')
Population=Data[['Persons']+['State']].groupby('State').sum()
Data_area=pds.sqldf('select a.State,a.Persons,b.Area from Population a join Area b on a.State=b.State')
Pop_list=Data_area['Persons'].tolist()
Area=Data_area['Area'].tolist()
Density=[]
for i in range(0,len(Pop_list)):
    Density.append(Pop_list[i]/Area[i])
    plt.subplot(111)
mu=np.mean(Density)
sigma=np.var(Density)
#plt.plot(Density,mlab.normpdf(Density, mu, sigma))
#pdf = stats.norm.pdf(Density, mu, sigma)
#plt.plot(Density,pdf)
plt.hist(Density)
plt.title('Histogram of Population Densities of different states')
plt.xlabel('Population Density')
plt.ylabel('Frequency')
plt.show()
```
