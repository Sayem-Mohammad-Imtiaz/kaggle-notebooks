```python
def find_value_counts(df):
    return df.value_counts()
def find_is_null(df):
    return df.isna().sum()
def find_unique(df):
    return df.unique()
def fill_na(df,value):
    df = df.fillna(value)
    return df
```

```python
df["rate"] = fill_na(df["rate"],0)
```

```python
df["rate"] = df["rate"].astype("str")
```

```python
def clean_rating(df):
    converted_list_1 = []
    converted_list_2 = []
    for i in range(len(df["rate"])):
         j = df["rate"][i][:3]
         converted_list_1.append(j)
    df["rate"]=pd.Series(converted_list_1)
    for i in df["rate"]:
        if i == "NEW" or i == "-":
            i = '0'
            converted_list_2.append(i)
        converted_list_2.append(i)
    df["rate"]=pd.Series(converted_list_2)
clean_rating(df)
```

```python
def split_multi_values(df):
    df_split = df.str.split(',',n=10, expand=True) 
    df_split.columns = [df.name+'_{}'.format(x+1) for x in df_split.columns]
    return df_split
```

```python
df_new = split_multi_values(df["rest_type"])
df["rest_type_1"] = df_new["rest_type_1"]
df["rest_type_2"] = df_new["rest_type_2"]
```

```python
df["rest_type_2"] = fill_na(df["rest_type_2"],"Not Available")
```

```python
df["approx_cost(for two people)"] = fill_na(df["approx_cost(for two people)"],"0")
```

```python
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].apply(lambda x: x.replace(',', ''))
```

```python
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].astype("int")
```

```python
def clean_approx_cost(df):
    converted_list_1 = []
    for i in df["approx_cost(for two people)"]:
        if i == 0:
            i = int(df["approx_cost(for two people)"].mean())
            converted_list_1.append(i)
        converted_list_1.append(i)
    df["approx_cost(for two people)"]=pd.Series(converted_list_1)
clean_approx_cost(df)
```

```python
df = df.drop(["phone","rest_type","dish_liked","cuisines","approx_cost(for two people)"],axis = 1)
```

```python
def change_binary(df):
    converted_list_1 = []
    for i in df:
        if i == "Yes":
            i = 1
            converted_list_1.append(i)
        else:
            i = 0
            converted_list_1.append(i)
    df=pd.Series(converted_list_1)
    return df
```

```python
df["online_order"] = change_binary(df["online_order"])
df["book_table"] = change_binary(df["book_table"])
```
