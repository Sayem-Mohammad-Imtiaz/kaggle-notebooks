
**Steps**
1. Rename "Metadata_Country_API_4_DS2_en_csv_v2_2710497" to "Metadata_Country" and "API_4_DS2_en_csv_v2_2710497" to "data"
2. Removed the first 4 rows from "data.csv" as they contains the "Data Source" and "Last Update date"
3. Rename all those column according to *BigQuery* naming convension. Adding "\_" in place of space and "\_" at the end of the year
4. Created a new project in *BigQuery* and upload those .csv files.
5. Use code to find missing "Country_Code"(I):
        SELECT DISTINCT DT.Country_Name 
        FROM `world-bank-07-08-2021.development_indicators.Metadata_Country` AS MD
        FULL OUTER JOIN `world-bank-07-08-2021.development_indicators.data` AS DT
        ON MD.Country_Code = DT.Country_Code
        WHERE MD.Country_Code IS NULL 
        OR DT.Country_Code IS NULL
   Country code for the country is "Not classified"
6. To find detalis about "Not classified":
        SELECT *
        FROM `world-bank-07-08-2021.development_indicators.data` 
        WHERE Country_Name = "Not classified"
   But all informations are null in this country. So it might be used as template or forget to enter the data

7. Another output is generated to find only 29 indicator(II) of 217 countries(III):
            SELECT *
            FROM `world-bank-07-08-2021.development_indicators.data` AS DT
            INNER JOIN 
                    (SELECT *
                    FROM `world-bank-07-08-2021.development_indicators.Metadata_Country`
                    WHERE Region is not null) AS MD
            ON DT.Country_Code=MD.Country_Code
            WHERE Indicator_Code="SL.UEM.TOTL.ZS" OR Indicator_Code="SH.DYN.0509" OR Indicator_Code="SE.XPD.PRIM.ZS" OR
                  Indicator_Code="SE.XPD.PRIM.PC.ZS" OR Indicator_Code="SE.PRM.UNER.ZS" OR Indicator_Code="SE.PRM.UNER.MA.ZS" OR
                  Indicator_Code="SE.PRM.UNER.FE.ZS" OR Indicator_Code="SE.PRM.TENR.MA" OR Indicator_Code="SE.PRM.TENR.FE" OR
                  Indicator_Code="SE.PRM.TENR" OR Indicator_Code="SE.PRM.TCHR.FE.ZS" OR Indicator_Code="SE.PRM.TCAQ.ZS" OR
                  Indicator_Code="SE.PRM.TCAQ.MA.ZS" OR Indicator_Code="SE.PRM.TCAQ.FE.ZS" OR Indicator_Code="SE.PRM.PRSL.ZS" OR
                  Indicator_Code="SE.PRM.PRSL.MA.ZS" OR Indicator_Code="SE.PRM.PRSL.FE.ZS" OR Indicator_Code="SE.PRM.OENR.ZS" OR
                  Indicator_Code="SE.PRM.OENR.MA.ZS" OR Indicator_Code="SE.PRM.OENR.FE.ZS" OR Indicator_Code="SE.PRM.NENR.MA" OR
                  Indicator_Code="SE.PRM.NENR.FE" OR Indicator_Code="SE.PRM.NENR"  OR Indicator_Code="SE.PRM.ENRL.TC.ZS" OR
                  Indicator_Code="SE.PRM.DURS" OR Indicator_Code="SE.PRM.CMPT.ZS" OR Indicator_Code="SE.PRM.CMPT.MA.ZS" OR
                  Indicator_Code="SE.PRM.CMPT.FE.ZS" OR Indicator_Code="SE.PRM.AGES" 
 
8. This output is save as .csv and downloaded and renamed as "result"
9. A new csv file "EduTime.csv" is created(IV). Use *=COUNTA(S2:X2)* to count data available in their primary education duration. And used *=IF(BT2=0,,SUM(S2:X2)/BT2)* to determine the mean.
10. A new csv file "bef edu.csv" is created(IV).
11. "bef edu.csv","EduTime.csv" and rest of the indicator form "result" are marged together to generate "Final WB upload.csv" file. 
12. As vlues for

            Adjusted net enrollment rate, primary (% of primary school age children)
            Adjusted net enrollment rate, primary, female (% of primary school age children)
            Adjusted net enrollment rate, primary, male (% of primary school age children)
are almose same to 
            School enrollment, primary (% net)
            School enrollment, primary, female (% net)
            School enrollment, primary, male (% net). 
So first 3 indicator are removed from the csv file.
13. Before computing each indicator in next stage, two filters are applied on enrollment and completion rate on that gender.
            a) *conditional formatting*, *Less than or equal to* 100
            b) *Filter by value* all other than *None*
14. Finally the three indicators are created by this:
            Dropout male % = School enrollment, primary, male (% net)-Primary completion rate, male (% of relevant age group)
            Dropout female % = School enrollment, primary, female (% net)-Primary completion rate, female (% of relevant age group)
            Dropout total % =School enrollment, primary, (% net)-Primary completion rate, total (% of relevant age group)
14. These three are aslo removed. As those also include improvement exam from precious exam.
            Primary completion rate, female (% of relevant age group)
            Primary completion rate, male (% of relevant age group)
            Primary completion rate, total (% of relevant age group)
15. To visualized only mean of availabel dropout is used. 
            dropout total=100- SUM(Persistence to last grade of primary, total (% of cohort))/COUNT(Persistence to last grade of primary, total (% of cohort))
    For all three indicator.
16. Then we created map with *Tableau public*.
