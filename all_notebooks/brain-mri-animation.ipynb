```python
from tqdm import tqdm
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import animation, rc
from pydicom import dcmread
rc('animation', html='jshtml')
```

```python
scans = !ls ../input/brain-mri-data/dicom-flair/dicom-flair/* -d
print(len(scans))
print(scans[0:3])
```

```python
def read_scan(path):
    fs = !ls -d {path}/*   
    slices = []
    for f in fs:
        ds = dcmread(f)
        data = ds.pixel_array
        num = int(ds.InstanceNumber)
        slices.append((num, data))
    
    slices.sort()
    slices = [s[1] for s in slices]
    return slices

def create_animation(ims):
    ims = ims
    fps = 30
    nSeconds = 5
    fig = plt.figure(figsize=(8,8))
    a = ims[0]
    im = plt.imshow(a)

    def animate_func(i):
        im.set_array(ims[i])
        return [im]

    anim = animation.FuncAnimation(fig, animate_func, frames = len(ims), interval = 1000//20)
    
    return anim
```

```python
ims=[]
s1 = scans[5:6]
for item in s1:
    ims += read_scan(item)
anim=create_animation(ims)
```
