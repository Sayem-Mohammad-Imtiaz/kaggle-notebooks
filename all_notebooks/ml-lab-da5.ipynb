```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv("../input/ml-da-5/place2.csv", header = None)
df=df.dropna()
df.head()
```

```python
#Converting the dataframe into an array of coordinates
arr=[]
for i in range (0, 300):
    x = df[0].values[i]
    y = df[0].values[i]
    arr.append([x,y])
```

```python
#Unsupervised Learning through KMeans Clustering
from sklearn.cluster import KMeans
X = np.array(arr)
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
kmeans.labels_
```

```python
#Calculating the score using internal cluster validation method - Sihouette score
from sklearn import metrics
labels = kmeans.labels_
metrics.silhouette_score(X, labels, metric = 'euclidean')
```

```python
#Displaying the results as mentioned in the question
Out = pd.DataFrame(labels, columns = ['Cluster_Label'])
Out
```
