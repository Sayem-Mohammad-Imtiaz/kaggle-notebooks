```python
import numpy as np
np.round(df['subject'].value_counts())
```

```python
np.round(df['subject'].value_counts(normalize=True),2)
```

```python
df['subject'].value_counts(normalize=False) ##If True then the object returned will contain the relative frequencies of the unique values
```

```python

df['year']= df['published_timestamp'].dt.year
subject_by_year = df.groupby('year')['subject'].value_counts().reset_index(level=0)
subject_by_year

```

```python
px.line(subject_by_year, x="year", y="subject",color= subject_by_year.index)
```

```python
levels = np.round(df['level'].value_counts(normalize=True),2)
levels
```

```python
px.histogram(df, x="level", title='Course Count by Level of Courses')
```

```python
levelss = df.groupby('year')['level'].value_counts().reset_index(level=0)
levelss
```

```python
px.line(levelss , x='year' , y="level" ,color=levelss.index)
```
