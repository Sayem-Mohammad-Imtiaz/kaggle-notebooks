### Importing Required Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="whitegrid")
```

```python
#For plotting the top 20 rider lets take a new dataframe consisting of the top 20 riders.
top_20 = df.iloc[:20]
top_20.shape
```

```python
#plot size
f, ax = plt.subplots(figsize=(16,9))

#plotting victories
sns.set_color_codes("pastel")
sns.barplot(x="victories", y="Name", data=top_20,
            label="1st place", color="b")

#plotting World Championships
sns.set_color_codes("muted")
sns.barplot(x="World Championships", y="Name", data=top_20,
            label="World Championships", color="b")

# Adding a legend and informative axis label
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, 100), ylabel="",
       xlabel="Top 20 MotoGP riders of all time.")
sns.despine(left=True, bottom=True)
```
