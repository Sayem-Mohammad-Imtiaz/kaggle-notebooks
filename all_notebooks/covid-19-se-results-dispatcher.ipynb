This page contains links to some SE results, this page will be updated frequently, as soon as we add new queries.

```python
from IPython.core.display import display, HTML

def display_links_tbl(queries):
    """
    Queries: list of tuple (query, link)
    """
    master_link = "https://www.kaggle.com/atmarouane/"

    table_html = "<ul>"
    for q, l in queries:
        link_id = master_link + l
        link = f"<a href='#' onclick=\"window.open('{link_id}');return false;\">{q}</a>"
        table_html += f"<li>{link}</li>"
    table_html += "</ul>"

    display(HTML(table_html))
```
