# 2. Correlation using Pearson's method (Pearson's Correlation)

***Why do we use this method?***

1. Used to find the magnitude/strength of the correlation between variables.
2. More accurate than that of scatter plot.

***Why is it better than scatter plot?***

1. Scatter plots are difficult to interpret.
2. If there is a lot of data, there is a higher probability of more than one scatter overlapping each other.
3. This makes the scatter plot difficult to plot as well for we would have to take care of its shortcomings, while on the other hand finding the correlation using a short and efficient algorithm is quite easy

***Why use Pearson's correlation?***

1. Because the relation between all of the variables is linear and Pearson's correlation works well with linear correlations.
2. Were the relations non linear, we would have used ****"Spearman's Rank Correlation"**** .

```python
def de_mean(x):
    x_bar = np.mean(x)
    return [x_i - x_bar for x_i in x]

def covariance(x, y):
    n = len(x)
    return np.dot(de_mean(x), de_mean(y)) / (n - 1)
```

```python
def correlation(x, y):
    std_x = np.std(x)
    std_y = np.std(y)
    if std_x > 0 and std_y > 0:
        return covariance(x, y) / (std_x * std_y)
        # we can also return covariance(x, y) / std_x / std_y
    else:
        return 0
```

```python
ms = df['math score']
rs = df['reading score']
ws = df['writing score']
```

```python
print('correlation between Maths score and reading score: ', correlation(ms, rs))
print('correlation between Maths score and writing score: ', correlation(ms, ws))
print('correlation between writing score and reading score: ', correlation(ws, rs))
```

***Conclusion***

1. Our prediction; that atleast 2 correlations would be quite more than 0.8 is correct and that the correlation between writing and reading scores would be more than 0.9 also is correct.
