<h1 style = "background:lightgreen;border:0"> Outlier Detection </h1>

**<font color = "darkred">Outlier :</font>**  An outlier is a data point in a data set that is distant from all other 
observations. A data point that lies outside the overall distribution of the dataset.

**<font color = "darkred">
Q1 --> </font>** First quarter. <br>
**<font color = "darkred">
Q3 --> </font>** Third quarter. <br> <br>
**<font color = "darkred">
IQR --> </font>** **<font color = "darkblue">Q3 - Q1</font>** **(interquartile distance)** <br> <br>
**<font color = "darkred">
lower bound --> </font>** **<font color = "darkblue">Q1 – 1.5</font>(IQR)** <br>
**<font color = "darkred">upper limit --></font>** 
**<font color = "darkblue">Q3 + 1.5</font>(IQR)**<br> <br>
**<font color = "darkred">
outlier --> </font>** **Values outside the lower and upper limits**

```python
# we write our function to find the outlier
def detect_outliers(df,features):
    outlier_indices = []
    
    for c in features:
        # 1st quartile
        Q1 = np.percentile(df[c],25)
        # 3st quartile
        Q3 = np.percentile(df[c],75)
        # IQR
        IQR = Q3 - Q1
        # Outlier Step
        outlier_step = IQR * 1.5
        # detect outlier and their indeces
        outlier_list_col = df[(df[c] < Q1 - outlier_step) | (df[c] > Q3 + outlier_step)].index
        # store indeces 
        outlier_indices.extend(outlier_list_col)
        
    outlier_indices = Counter(outlier_indices)
    """
    ## Usually at least 2 outliers in a row we drop.
    ## if there are two discrete values
    """
    multiple_outliers = list(i for i, v in outlier_indices.items() if v > 2) 
    
    return multiple_outliers
```

<p style = "text-indent:25px">Line plot in Plotly is much accessible and  illustrious annexation to plotly which manage variety of types of data, and assemble easy-to-style statistic.</p>

<hr />

<strong style = "color:darkblue">Graphic Creation Stages(style1) :</strong>
<p style = "color:darkred"> 1 -> Creating Traces :(Trace1,Trace2,...)</p>
trace1 = go.Scatter(the graphic we use)(
    <li style = "text-indent:80px;">x : x-axis</li>
    <li style = "text-indent:80px;">y : y-axis</li>
    <li style = "text-indent:80px;">mode</li>
    <li style = "text-indent:80px;">name</li>
    <li style = "text-indent:80px;">marker</li>
    <li style = "text-indent:80px;">text )</li> <br>
    
<p style = "color:darkred">2 -> Data :</p>
 <p>data = [Trace1,Trace2]</p> 
<p style = "color:darkred">3 -> layout : outer part of the chart.</p>
<p> layout = dict(title : chart title,
    xaxis = dict(title = label of x axis,ticklen = length of x axis ticks, zeroline = showing zero line or not),yaxis = dict(title: label of y axis))

<p style = "color:darkred">4 -> fig : it includes data and layout.</p>
<p>fig = dict(data = data, layout = layout)</p> 
<p style = "color:darkred" >5 -> iplot() : Draws the generated chart.</p>
<p>iplot(fig)</p>


```python
### Social support and Freedom to make life choices vs Ladder score of Top 30 Countries

# top 30 countries in the world ranking
top_countries_onehundred = world_happiness_report_2020.iloc[:30,:]
"""
rgba(red,green,blue,alpha) = different color rendering.
You can create the color you want by changing the values.

"""
# Creating trace1(Social Support)
trace1 = go.Scatter(
                    x = top_countries_onehundred["Ladder score"], # x-axis
                    y = top_countries_onehundred["Social support"], # y-axis
                    mode = "lines", # how to draw
                    name = "Social support",
                    marker = dict(color = 'rgba(255, 20, 30, 1)'),# color of the lines
                    text= top_countries_onehundred["Country name"]) # text on hover
# Creating trace2(Freedom to make life choices)
trace2 = go.Scatter(
                    x = top_countries_onehundred["Ladder score"],# x-axis
                    y = top_countries_onehundred["Freedom to make life choices"], # y-axis
                    mode = "lines+markers", # how to draw
                    name = "Freedom to make life choices",
                    marker = dict(color = 'rgba(15, 99, 36, 0.8)'),
                    text= top_countries_onehundred["Country name"])

# we combine our two traces
data = [trace1, trace2]
# outer part of the chart
layout = dict(title = 'Social support and Freedom to make life choices vs Ladder score of Top 30 Countries',
              xaxis= dict(title= 'Ladder score',ticklen= 15,zeroline= False)
             )
# We combine our data and layout
# create figure
fig = dict(data = data, layout = layout)
iplot(fig) # draw the chart
```

```python
"""
We will create an example using typ 3.
We want to see the happiness rate of the regions
"""

# We put our singular regions on a list
regions =list(world_happiness_report_2020["Regional indicator"].unique()) 
# happiness rates of regions
regions_happiness_rate = []
"""
To find the happiness rate of the region;
We collect the happiness rate of the countries in that region.
and divide by the number of countries in that region.

example: 
i = western Europe
x = all countries in this region.
happiness_rate = sum of happiness rates / x
"""
for i in regions:
    x = world_happiness_report_2020[world_happiness_report_2020["Regional indicator"] == i]
    happiness_rate = sum(x["Ladder score"]) / len(x)
    regions_happiness_rate.append(happiness_rate) # add list

# figure
# We create a figure and do it for its operations.
fig = {
  "data": [
    {
      "values": regions_happiness_rate,
      "labels": regions, # region names on the side
      "domain": {"x": [0, .8]}, # chart size
      "name": "Happiness Rate",
      "hoverinfo":"label+percent+name", # text on hover
      "hole": .3, # the size of the hole in the middle
      "type": "pie" # chart type
    },],
  "layout": {
        "title":"Happiness Rates by Region", # chart title
        "annotations": [ # regulation of chart
            { "font": { "size": 13},
              "showarrow": True, # show arrow on
              "text": "Happiness Rate", # arrow text
                "x": 0.2, # x-axis
                "y": 0.9 # y-axis
            },
        ]
    }
}
iplot(fig)
```

<h2 style = "background:lightgreen;border:0">Bubble Plot</h2>
<p>
A bubble chart is a scatter plot in which a third dimension of the data is shown through the size of markers.New dimensions are gained by giving the color and size properties.
</p>

<strong>Parameters:</strong> The description of some main parameters are given below:
<ul>    
<li><strong>x, y : </strong> data for the chart.(inputs)</li>
<li><strong>mode :</strong> type of line on chart.(lines, markers, vs.)</li>
<li><strong>name :</strong>the name of the line.</li>
<li><strong>hoverinfo :</strong>the text that will appear when mouse is over the graphic</li>    
<li><strong>marker :</strong>lines have properties.(color, vs.)</li> 
    </ul>

```python
"""
We will look at the happiness 
rates of the regions and the number of countries in those regions in a single graph.

y = happiness rate
x = region names
color = Healthy life expectancy
size = number of countries
"""
# regions
regions =list(world_happiness_report_2020["Regional indicator"].unique()) 
# happiness rates of regions
regions_happiness_rate = []
# Healthy life expectancy
healty_rate = []
# region_names
region_names = []
# number of countries
number_countries = [] 

"""
we have countries' happiness rate and healthy life expectancy.
We will calculate the regions using them.

i = region
x = all countries in that region
happiness_rate = happiness rate of the region.
healty = average healthy life expectancy of the region
region_ names = region names
number_countries = the number of countries in that region
"""
for i in regions:
    x = world_happiness_report_2020[world_happiness_report_2020["Regional indicator"] == i]
    happiness_rate = sum(x["Ladder score"]) / len(x)
    healty = sum(x["Healthy life expectancy"] / len(x))
    healty_rate.append(healty) # healty rate add
    regions_happiness_rate.append(happiness_rate) # add list
    region_names.append(i) # region names add
    number_countries.append(len(x))  # number of countries add

data = [
    {
        'y': regions_happiness_rate,
        'x': region_names,
        'mode': 'markers',
        'name': 'Happiness Rate',
        "hoverinfo":"x+y+name", # text on hover
        'marker': {
            'color': healty_rate,# health rate of the region
            'size': number_countries, # by the number of countries
            'showscale': True #scale on the side
        },
            
    }
]
iplot(data)
```

```python
# create trace 1 that is 3d scatter
trace1 = go.Scatter3d(
    x=world_happiness_report_2020["Country name"],
    y=world_happiness_report_2020["Generosity"],
    z=world_happiness_report_2020["Social support"],
    mode='markers',
    marker=dict(
        size=6,
        color= world_happiness_report_2020["Overall rank"],  # set color to an array/list of desired values      
        
    )
)

data = [trace1]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0  
    )
    
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

<a id ='22' ></a>
<h2 style = "background:lightgreen;border:0">Multiple Subplots</h2>
<p>It allows us to see many sub graphs in a single shape..</p>

<strong>Parameters:</strong> The description of some main parameters are given below:
<ul>    
<li><strong>x, y : </strong> data for the chart.(inputs).</li>
<li><strong>name :</strong> the name of the lines.</li>    
<li><strong>xaxis :</strong> x-axis of the small chart.</li> 
<li><strong>yaxis :</strong> y-axis of the small chart.</li>     
    </ul>

```python
# create trace1 
trace1 = go.Scatter(
    x=world_happiness_report_2020["Overall rank"], 
    y=world_happiness_report_2020["Social support"],
    name = "Social support"
)
# create trace2 
trace2 = go.Scatter(
    x=world_happiness_report_2020["Overall rank"],
    y=world_happiness_report_2020["Logged GDP per capita"],
    xaxis='x2',# x-axis of bottom chart
    yaxis='y2',# y-axis of bottom chart
    name = "Logged GDP per capita" # name of bottom chart
)
trace3 = go.Scatter(
    x=world_happiness_report_2020["Overall rank"],
    y=world_happiness_report_2020["Generosity"],
    xaxis='x3',# x-axis of bottom chart
    yaxis='y3',# y-axis of bottom chart
    name = "income" # name of bottom chart
)
trace4 = go.Scatter(
    x=world_happiness_report_2020["Overall rank"],
    y=world_happiness_report_2020["Healthy life expectancy"],
    xaxis='x4',# x-axis of bottom chart
    yaxis='y4',# y-axis of bottom chart
    name = "Healthy life expectancy" # name of bottom chart
)
data = [trace1, trace2, trace3, trace4]

layout = go.Layout(
    xaxis=dict(
        domain=[0, 0.45] # We write the properties of my axes.
    ),
    yaxis=dict(
        domain=[0, 0.45]
    ),
    xaxis2=dict(
        domain=[0.55, 1]
    ),
    xaxis3=dict(
        domain=[0, 0.45],
        anchor='y3'
    ),
    xaxis4=dict(
        domain=[0.55, 1],
        anchor='y4'
    ),
    yaxis2=dict(
        domain=[0, 0.45],
        anchor='x2'
    ),
    yaxis3=dict(
        domain=[0.55, 1]
    ),
    yaxis4=dict(
        domain=[0.55, 1],
        anchor='x4'
    ),
    title = 'Social support, Logged GDP per capita, Generosity and Healthy life expectancy VS world happiness rankings'
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

<h2 style = "background:lightgreen;border:0">Choropleth Map</h2>
<p>we use our data to view maps..</p>

<strong>Parameters:</strong> The description of some main parameters are given below:
<ul>    
<li><strong>locations : </strong> data to be given according to the selected mode(we give the locations).</li>
<li><strong>locationmode :</strong> We tell you how to draw the map(‘ISO-3’, ‘USA-states’, or ‘country names’).</li>    
<li><strong>projection :</strong> How will the map projections be.There are many types.(example : 'kavrayskiy7', 'miller', 'robinson', 'eckert4', 'azimuthal equal area').</li> 
<li><strong>hover_data :</strong>  information to be given when mouse over the country.</li>  
<li><strong>hover_name :</strong> We can call it the title of hover data.</li>
<li><strong>color :</strong> Coloring the map..</li>  
<li><strong>scope :</strong> viewpoint on the map.</li>    
<li><strong>color_continuous_scale :</strong> We can call it the form of coloring..</li>    
    </ul>

```python
# we import our library
import plotly.express as px

# We choose our top 50 countries
happy = world_happiness_report_2020.head(50)

# information to be given when mouse over the country
hover_data = world_happiness_report_2020[["Overall rank","Regional indicator","Ladder score","Logged GDP per capita","Social support","Generosity"]]

fig = px.choropleth(happy,
                    locations=happy["Country name"],
                    locationmode="country names",
                    projection="natural earth",
                    hover_data=hover_data,
                    hover_name=happy["Country name"],
                    color="Overall rank",
                    color_continuous_scale=px.colors.sequential.Rainbow,
                    scope="world")
fig.show()
```
