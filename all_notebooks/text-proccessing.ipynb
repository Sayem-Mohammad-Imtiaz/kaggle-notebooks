```python
import nltk
import string
import numpy as np
import pandas as pd
from sklearn.svm import SVC
from nltk.corpus import stopwords
from sklearn.decomposition import TruncatedSVD
from nltk.tokenize import word_tokenize
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
train = pd.read_csv('../input/is-humor-humor-controversy-data-set/train.csv')
stopwords = set(stopwords.words('english'))
train.fillna(0, inplace=True)
train.head(1)
```

```python
X = []
for text in train['text']:
    new_text = []
    for word in word_tokenize(text):
        word = word.lower()
        if word not in stopwords and word not in string.punctuation:
            new_text.append(word)
    result = ' '.join(new_text)
    X.append(result)
```

```python
Y1 = train['is_humor']
Y2 = train['humor_controversy']
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1, 2))
X = vectorizer.fit_transform(X)
clf = TruncatedSVD(1000)
X = clf.fit_transform(X)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y1, test_size=0.1, shuffle=0.5)
print(X.shape)
```

```python
model_svc = SVC(kernel='linear')
model_svc.fit(X_train, Y_train)
```

```python
Y_pre = model_svc.predict(X_test)
```

```python
print(accuracy_score(Y_pre, Y_test)*100,'\n')
print(confusion_matrix(Y_pre, Y_test))
```
