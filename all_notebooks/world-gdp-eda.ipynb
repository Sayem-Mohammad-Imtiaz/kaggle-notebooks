```python
east_africa = ['Kenya', 'Uganda', 'Tanzania', 'Rwanda', 'Burundi', 'South Sudan']

east_africa_gdp = gdp_csv[gdp_csv['Country Name'].isin(east_africa)]
print(len(east_africa_gdp))
east_africa_gdp['Country Code'].unique()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['figure.figsize'] = (14,8)
sns.set()
```

```python
fig, ax = plt.subplots()

gdp = east_africa_gdp
for country in east_africa:
    data = gdp[gdp['Country Name'] == country]
    sns.lineplot(x='Year', y='Value', data=data, ax=ax)
ax.legend(east_africa)
ax.set_title('GDP of East African countries from {:.0f} to {:.0f}'.format(
    gdp.Year.min(), gdp.Year.max()
))
plt.show()
```

```python
fig, ax = plt.subplots()

gdp = east_africa_gdp
gdp = gdp[gdp['Year'] <= 1980]
for country in east_africa:
    data = gdp[gdp['Country Name'] == country]
    sns.lineplot(x='Year', y='Value', data=data, ax=ax)
ax.legend(east_africa)
ax.set_title('GDP of East African countries from {:.0f} to {:.0f}'.format(
    gdp.Year.min(), gdp.Year.max()
))
plt.show()
```

```python
fig, ax = plt.subplots()

gdp = east_africa_gdp
gdp = gdp[gdp['Year'].isin(list(range(1980, 2001)))]
for country in east_africa:
    data = gdp[gdp['Country Name'] == country]
    sns.lineplot(x='Year', y='Value', data=data, ax=ax)
ax.legend(east_africa)
ax.set_title('GDP of East African countries from {:.0f} to {:.0f}'.format(
    gdp.Year.min(), gdp.Year.max()
))
plt.show()
```

```python
fig, ax = plt.subplots()

gdp = east_africa_gdp
gdp = gdp[gdp['Year'] >= 2000]
for country in east_africa:
    data = gdp[gdp['Country Name'] == country]
    sns.lineplot(x='Year', y='Value', data=data, ax=ax)
ax.legend(east_africa)
ax.set_title('GDP of East African countries from {:.0f} to {:.0f}'.format(
    gdp.Year.min(), gdp.Year.max()
))
plt.show()
```
