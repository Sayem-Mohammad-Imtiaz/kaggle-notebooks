```python
text=data['comment_text']
```

```python
text = text.str.replace("[^a-zA-Z#]", " ")
```

```python
import nltk
from nltk.corpus import stopwords
pre1=[]
for i in text:
    i=str(i)
    cont=i.split()
    stopword = stopwords.words('english')
    cont = [cont for cont in cont if cont not in stopword]
    new_cont = ' '.join(cont)
    pre1.append(new_cont)
```

```python
#소문자로 변경
pre2=[]
for i in pre1:
    sentence=i.lower()
    pre2.append(sentence)
```

```python
# lemmatize

n=WordNetLemmatizer()
pre3=[]
for i in pre2:
    words=word_tokenize(i)
    lem_word=[]
    for j in words:
        lemm=n.lemmatize(j)
        lem_word.append(lemm)
    ps_tg=pos_tag(lem_word)
    pre3.append(ps_tg)
```

```python
#명동형부사추출
pre4=[]
for i in pre3:
    a=[]
    for j in i:
        if (j[1]=='NN' or j[1]=='NNP' or j[1]=='NNS' or j[1]=='NNPS' or j[1]=='VB'or j[1]=='VBD'or j[1]=='VBG'or j[1]=='VBN'or j[1]=='VBP'or j[1]=='VBZ'or j[1]=='VH'or j[1]=='VHD'or j[1]=='VHG'or j[1]=='VHN'or j[1]=='VHP'or j[1]=='VHZ'or j[1]=='VV'or j[1]=='VVD'or j[1]=='VVG'or j[1]=='VVN'or j[1]=='VVP'or j[1]=='VVZ'or j[1]=='JJ'or j[1]=='JJR'or j[1]=='JJS'or j[1]=='RB'or j[1]=='RBR'or j[1]=='RBS'):
            a.append(j[0])
    pre4.append(a)
```

```python
pp=[]
for i in pre4:
    pp+=i
```

```python
ppp=[]
for i in pp:
    if len(i)>1:
        ppp.append(i)
```

```python
from collections import Counter

count = Counter(ppp)
count
```

```python
words = dict(count.most_common())
words
```
