```python
data=pd.read_csv('/kaggle/input/heart-disease-cleveland-uci/heart_cleveland_upload.csv')
data.tail()
```

```python
x=data.drop('condition',axis=1).copy()
y=data['condition'].copy()
x.head(),y.head()
```

```python
cp1=pd.get_dummies(data['cp'],drop_first=True)
cp1.tail()
```

```python
slp1=pd.get_dummies(data['slope'],drop_first=True)
cp1.tail()
```

```python
rg1=pd.get_dummies(data['restecg'],drop_first=True)
cp1.tail()
```

```python
t1=pd.get_dummies(data['thal'],drop_first=True)
cp1.tail()
```

```python
data=pd.concat([data,cp1,slp1,rg1,t1],axis=1)
data.head()
```

```python
import math
data.isnull().sum()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=100)
model=DecisionTreeClassifier(ccp_alpha=.016,max_depth=20,criterion='entropy',min_samples_leaf=10)
model=model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix
confusion_matrix(y_test,y_pred)
plot_confusion_matrix(model,x_test,y_test,display_labels=['no HD','has HD'])
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(15,7.5))
plot_tree(model,filled=True,rounded=True,class_names=['no hd','hd'],feature_names=x.columns)
```
