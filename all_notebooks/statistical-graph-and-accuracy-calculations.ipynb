```python
# Following Notebook reflects the use of statistical graph plots on logical gates dataset.
# Along with plots the accuracy score is calculated with the K nearest neighbors algorithm, here to define the plot we have used seaborn to set the style and context of the plot.
# here accuracy score is calculated on AND gate:

import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score

df=pd.read_csv("../input/logical-gate-dataset/logical_gate.csv")
print(df)

sns.set_context("notebook",font_scale=1.5)
sns.set_style("ticks")
sns.lmplot('Value_A','Value_B',data=df,fit_reg=True,scatter=True,hue='AND_Gate')
plt.show()

X=df[['Value_A','Value_B']]
Y=df['AND_Gate']

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.33,random_state=4)
knn=KNeighborsClassifier()
knn.fit(X_train,Y_train)
print("The Accuracy Score is {}".format(knn.score(X_train,Y_train)))
pred=knn.predict(X_test)
print(pred)
print(Y_test)
```
