# Selection of learing algorythm

```python
#I've decided to use Gradient Boosting Regressor as my first mode: it's a simple but effective algorythm.
```

```python
dftest = df[df['Month']==12]
dftest
```

```python
X = df1.iloc[:,0:4].values
Y = df1.iloc[:,4].values
```

```python
X_train, X_test,Y_train, Y_test = train_test_split (X, Y, test_size=0.2, random_state=7)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
def models (X_train,Y_train):
    from sklearn.ensemble import GradientBoostingRegressor
    xgbm = GradientBoostingRegressor(n_estimators=2000, 
                                      learning_rate =0.2, max_depth =1, random_state=5)
    xgbm.fit(X_train,Y_train)
    print ('XGB Training Accuracy:' , xgbm.score(X_train,Y_train))
    return xgbm
    
    
```

```python
model = models(X_train,Y_train)
```

```python
#Model has a 94% accuracy. In my humble opinion, it's quite an effective model, which does not need tuning.
```

# Costs predictions

```python
Xt = dftest.iloc[:,0:4].values
Xt = sc.transform(Xt)
```

```python
dftest['costs'] = model.predict(Xt)
```
