```python
import numpy as np
import pandas as pd
```

```python
#1
#read file
df = pd.read_csv("../input/hotel-booking/hotel_booking.csv")
```

```python
#2
#check data
print(df.info())
print(df.describe())
```

```python
#3
#number of rows
df.shape[0]
#Or len(df.index)
```

```python
#4
#number of missing data
print(df.isnull().values.any())
print(df.isna().sum().sum())
```

```python
#5
#remove company column
df.drop(["company"],axis=1,inplace=True)
```

```python
#6
#repetitive countries
df['country'].value_counts()[0:5]
```

```python
#7
#Most adr
maxADR=df["adr"].max()
num=df["adr"].idxmax()
"{a} - {b}".format(a=maxADR,b=df.iloc[num]["name"])
```

```python
#8
#Average adr
round(df['adr'].mean(), 2)
```

```python
#9
#Average stays night
df["stays_night"] = df["stays_in_week_nights"] + df["stays_in_weekend_nights"]
round(df['stays_night'].mean(), 2)
```

```python
#10
#determine the people who have the most special requests
special_request = df[df["total_of_special_requests"] == 5]
special_request[["name","email"]]
```

```python
#11
#Frequent families
fn=lambda n:str(n).split()
df["family_name"]=df["name"].apply(fn)
print(df["family_name"].value_counts()[0:5])
```

```python
#12
#People with the most immature companions
df["not_adult"] = df["babies"] + df["children"]
df[["name","not_adult"]].sort_values("not_adult", ascending=False)[0:2]
```

```python
#13
#Area code of the most booked 
df["phone-number"].apply(lambda p : p[0:3]).value_counts()[0:3]
```
