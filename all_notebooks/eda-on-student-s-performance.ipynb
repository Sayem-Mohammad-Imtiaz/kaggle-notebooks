# Student Performance In Exams                                                                            

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
studata = pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
```

```python
studata.head()
```

In this Data set we have:-

Numerical Variables- Math score, Reading score and Writing score.

Categorical Variables - Gender, Race/ethnicity, Parental level of education, Lunch and Test preparation course.



```python
studata.info()
```

There are no missing ( null ) values in this dataframe. 

```python
studata['total score']=(studata['math score'] + studata['reading score'] + studata['writing score'])/3
```

Let's enter a total score column, so that we can analyze a student performance as a whole.

```python
studata.head()
```

```python
studata.describe()
```

```python
sns.heatmap(studata.corr())
```

# Exploring the scores of the students.

```python
plt.figure(figsize=(20, 10))
sns.countplot(studata['math score'], palette = 'Set1').set_title('Math Score')
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(studata['reading score'], palette = 'Set2').set_title('Math Score')
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(studata['writing score'], palette = 'Set3').set_title('Math Score')
```

Based on the graphs, we can say that maximum number of students have scored between 60-80 in all the three subjects.

# Effect of Gender on Student's Performnce

```python
studata.groupby(['gender']).mean()
```

```python
studata['gender'].unique()
```

```python
plt.figure(figsize=(5, 5))
plt.title('Gender')
studata['gender'].value_counts().plot.pie(autopct="%1.1f%%")
```

Proportion of male and female are almost same

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.boxplot(x = 'gender', y = 'math score', data = studata).set_title('GENDER VS MATH SCORE')

plt.subplot(1,3,2)
sns.boxplot(x = 'gender', y = 'reading score', data = studata).set_title('GENDER VS READING SCORE')

plt.subplot(1,3,3)
sns.boxplot(x = 'gender', y = 'writing score', data = studata).set_title('GENDER VS WRITING SCORE')
```

```python
sns.barplot(x='total score',y='gender',data=studata).set_title('GENDER VS TOTAL SCORE')
```

If we analyze the overall performance of the students then, female candidates have scored higher marks than the male candidates.

Subjectwise analysis:-
The male candidates scored better in maths compared to female candidates.
The female candidates scored better in writing and reading compared to male candidates.

# Effect of Lunch on Student's Performnce

```python
studata.groupby(['lunch']).mean()
```

```python
studata['lunch'].unique()
```

```python
plt.figure(figsize=(5, 5))
plt.title('Lunch')
studata['lunch'].value_counts().plot.pie(autopct="%1.1f%%")
```

Large proportion of the students have standard lunch.

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.violinplot(x = 'lunch' , y = 'math score' , data = studata).set_title('LUNCH VS MATH SCORE')

plt.subplot(1,3,2)
sns.violinplot(x ='lunch', y = 'reading score' , data = studata).set_title('LUNCH VS READING SCORE')

plt.subplot(1,3,3)
sns.violinplot(x = 'lunch', y = 'writing score' , data = studata).set_title('LUNCH VS WRITING SCORE')
```

```python
sns.barplot(x='total score',y='lunch',data=studata).set_title('LUNCH VS TOTAL SCORE')
```

Based on the analysis, students who have opted for standard lunch seem to perform better than the students who have opted free/reduced lunch.

This trend also follows for student's subjectwise performance.

# Effect of parental level of education on Student's Performnce

```python
studata.groupby(['parental level of education']).mean()
```

```python
studata['parental level of education'].unique()
```

```python
plt.figure(figsize=(5, 5))
plt.title('Parental level of education')
studata['parental level of education'].value_counts().plot.pie(autopct="%1.1f%%")
```

Highest proportion of students with parentel level of Education is 'Some college', 'associate's degreee' and 'high school'.

```python
plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.barplot(x = 'parental level of education' , y = 'math score' , data = studata).set_title('PARENTAL LEVEL OF EDUCATION VS MATH SCORE')
plt.xticks(rotation = 90)

plt.subplot(1,3,2)
sns.barplot(x ='parental level of education', y = 'reading score' , data = studata).set_title('PARENTAL LEVEL OF EDUCATION VS READING SCORE')
plt.xticks(rotation = 90)

plt.subplot(1,3,3)
sns.barplot(x = 'parental level of education', y = 'writing score' , data = studata).set_title('PARENTAL LEVEL OF EDUCATION VS WRITING SCORE')
plt.xticks(rotation = 90)

plt.tight_layout()
```

```python
sns.barplot(x='total score',y='parental level of education',data=studata).set_title('PARENTAL LEVEL OF EDUCATION VS WRITING SCORE VS TOTAL SCORE')
```

Based on the analysis:-
    
Student's whose parents have master's degree have the highest score in all subjects and performed well.

Student's whose parents have bachelor's degree or associate degree performed reasonably well in all subjects.

Student's whose parents level of education is college or school level comparitively have not performed well compared to other students.

# Effect of Race/Ethnicity on Student's Performnce

```python
studata.groupby(['race/ethnicity']).mean()
```

```python
studata['race/ethnicity'].unique()
```

```python
sns.barplot(x='total score',y='race/ethnicity',data=studata).set_title('RACE/ETHNICITY VS TOTAL SCORE')
```

Based on the analysis:-
    
Students who belong to group E scored the highest scores compared to other groups.

Students who belong to group A scored the lowest scores compared to other groups.
