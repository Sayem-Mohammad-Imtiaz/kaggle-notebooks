```python
g = sns.catplot(
    data=df, kind="bar",
    x="cuisine", y="base_price", hue="center_type",
    ci="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "Base price ($)")
```

```python
g = sns.catplot(
    data=df, kind="bar",
    x="cuisine", y="checkout_price", hue="center_type",
    ci="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "checkout price ($)")
g.legend.set_title("")
```
