I wanted to see how often a team hits the over during the past 10 seasons. This will get us the count of how many time each time hit the over.

```python
import pandas as pd

teams = ['CHI', 'GNB', 'CAR', 'CLE', 'LAR', 'TEN', 'ARI', 'DET', 'DAL', 'NYG', 'JAX', 'KAN', 'MIA', 'BAL', 'MIN', 'ATL', 'NWE', 'PIT', 'NYJ', 'BUF', 'PHI', 'WAS', 'LAC', 'IND', 'SEA', 'CIN', 'TAM', 'SFO', 'NOR', 'HOU', 'OAK', 'DEN']
overs = []
for i in teams:
	total_overs = 0
	for j in range(2010, 2020):
		df = pd.read_csv(f'../input/nfl-game-stats-20102019/NFL_Game_Stats/game_stats_{j}.csv')

	
		if(j < 2015):
			if(i == 'LAR'):
				i = 'STL'
			elif(i == 'LAC'):
				i = 'SDG'
		elif(j < 2016):
			if(i == 'LAC'):
				i = 'SDG'

		team_h = df.loc[df['HomeTeam'] == i]
		overs_h = team_h.loc[df['Result'] == 1]
		team_a = df.loc[df['AwayTeam'] == i]
		overs_a = team_a.loc[df['Result'] == 1]
		num_overs = (overs_h['Result'].count() + overs_a['Result'].count())

		total_overs += num_overs

	overs.append(total_overs)
```

Now let's make a graph to visualize how often each time hits the over using a bar graph using matplotlib.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [20, 10]

labels = teams
vals = overs

x = np.arange(len(labels))  # the label locations
width = 0.5  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x, vals, width, label='Total Overs')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Total Overs')
ax.set_title('Total Overs by Team over the last 10 Years')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)


plt.show()
```

Looks like the New Orleans Saints have hit the over the most over the past 10 seasons. Who Dat! Of course I shouldn't really be celebrating because it pretty much shows how bad our defense has been over those seasons.
