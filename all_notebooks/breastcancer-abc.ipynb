```python
X=df
y=df['diagnosis']
X_train, x_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=10)
```

```python
cols=df.columns.drop('diagnosis')
fromula='diagnosis ~' +  '+'.join(cols)
print(fromula,'\n')
```

```python
model=smf.glm(formula=fromula,data=X_train, family=sm.families.Binomial())
logistic_fit=model.fit()

print(logistic_fit.summary())
```

```python
predict=logistic_fit.predict(x_test)
predict[1:6]
```

```python
prediction_nominal=['M' if x<0.5 else 'B' for x in predict ]
prediction_nominal[1:6]
```

```python
print(classification_report(y_test, prediction_nominal, digits=3))

cfm=confusion_matrix(y_test, prediction_nominal)
trueNegative=cfm[0][0]
falsePossitive=cfm[0][1]
false_negative=cfm[1][0]
truePositive=cfm[1][1]

print("Confudion Matrix: \n",cfm,'\n')
print("true negative", trueNegative)
print("False Positive", falsePossitive)
print("false Negative", false_negative)
print("True Positive", truePositive)

print("correct prediction", 
      round((trueNegative+truePositive)/len(prediction_nominal) *100, 1),'%')
```
