```python
df = pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
df.head()

```

```python
drop_A=df.index[df["Glucose"] == 0].tolist()
drop_B=df.index[df["BloodPressure"] == 0].tolist()
drop_C=df.index[df["SkinThickness"] == 0].tolist()
drop_D=df.index[df["Insulin"] == 0].tolist()
drop_E=df.index[df["BMI"] == 0].tolist()
drop_F=df.index[df["Age"] == 0].tolist()
drop=drop_A+drop_B+drop_C+drop_D+drop_E+drop_F
df=df.drop(df.index[drop])
```
