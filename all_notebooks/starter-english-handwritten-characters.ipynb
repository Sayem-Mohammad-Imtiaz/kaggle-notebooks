## Introduction

This dataset contains 3,410 handwritten English Characters that include 0-9, A-Z and a-z. This is a multiclass classification as each image is mapped to only one label from 62 such labels.

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing Data

For each image, we have it's class mentioned in `/kaggle/input/english-handwritten-characters/english.csv`. The `label` column is categorical in nature and `image` column has string datatype.

```python
df = pd.read_csv(
    "/kaggle/input/english-handwritten-characters-dataset/english.csv",
    dtype={"image": pd.StringDtype(), "label": pd.CategoricalDtype()}
)
```

Let's explore some images

```python
img = mpimg.imread("/kaggle/input/english-handwritten-characters-dataset/" + df.at[473, "image"])
plt.imshow(img)
```
