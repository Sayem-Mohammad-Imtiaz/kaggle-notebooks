```python
import pandas as pd 
from sklearn.preprocessing import MinMaxScaler
import numpy as np # linear algebra
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
```

<a id='1'></a><br>
# 1.Exploring Data

```python
car_data=pd.read_csv('../input/vehicle-dataset-from-cardekho/car data.csv')
```

```python
car_data.head()
```

```python
car_data.info()
```

<a id='2'></a><br>
# 2.Feature Engineering

```python
car_data['Transmission']=[ 0 if i=='Manual'  else 1  for i in car_data['Transmission']] #manuel 0 # auto 1 
car_data['Fuel_Type']=[ 0 if i=='Petrol'  else 1  for i in car_data['Fuel_Type']] #manuel 0 # auto 1 
```

```python
car_data.head()
```

<a id='3'></a><br>
# 3.Prepare Data

```python
x=car_data.loc[:,['Year','Selling_Price','Kms_Driven','Fuel_Type','Transmission']].values
y=car_data['Present_Price'].values.reshape(-1,1)
```

<a id='4'></a><br>
# 4.Normalization

```python
# scaling 
scaler = MinMaxScaler(feature_range=(0, 1))
x = scaler.fit_transform(x)
```

<a id='5'></a><br>
# 5.Train-Test Split

```python
# Train Test Split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.10,random_state=42)
```

<a id='6'></a><br>
# 6.Multiple Linear Regression 83% accuracy
y=b0+b1*x1+b2*x2+b3*x3+b4*x4+b5*x5

```python
# multiple linear regression
mlr=LinearRegression()
mlr.fit(x_train,y_train)
```

```python
print('b0',mlr.intercept_)
print('b1,b2,b3,b4,b5',mlr.coef_)
```

```python
# prediction
y_head=mlr.predict(x_test) 
```

```python
# r square with linear regression

from sklearn.metrics import r2_score

print('r2 score:',r2_score(y_test,y_head))
```

```python
# k kold cross validation 
scores = cross_val_score(mlr, x_train, y_train, scoring='r2', cv=5)
print('avarage accuracy:',np.mean(scores))
```

<a id='7'></a><br>
# 7.Random Forest Regression 95% 

```python
from sklearn.ensemble import RandomForestRegressor
random_forest_reg=RandomForestRegressor(n_estimators=100,random_state=42)
random_forest_reg.fit(x_train,y_train)
```

```python
# Predict  
y_head=random_forest_reg.predict(x_test)
```

```python
# r square
from sklearn.metrics import r2_score

print('r2 score:',r2_score(y_test,y_head))
```

# 8.Decision Tree Regression 91%

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg=DecisionTreeRegressor(random_state=42) # random state
tree_reg.fit(x_train,y_train)
```

```python
# Prediction
y_head=tree_reg.predict(x_test)
```
