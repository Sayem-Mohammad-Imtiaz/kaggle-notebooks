```python
import pandas as pd # data processing
```

Importing the data file

```python
df = pd.read_csv('/kaggle/input/data-cleaning/employee_data.csv')
```

```python
#General Info
df.info()
```

```python
# Numeric columns statistics
df.describe()
```

```python
#Coiunting NULLs in each column
df.isnull().sum()
```

#Dropping column "number" and "tax_file_no" , as both are not required

```python
df.drop(columns= ['number','tax_file_no'],inplace = True,axis=1)
```

```python
df.dropna(axis='index', thresh=2, inplace=True)
df.info()
```

```python
# Set a unknown category for missing genders.
df['gender'].fillna('U', inplace=True)
print(df['gender'].value_counts())
```

```python
df['employment_status'].fillna('Unknown', inplace=True)
print(df['employment_status'].value_counts())
```

```python
#Creating dummy variables
df = pd.get_dummies(df,columns = ['employment_status','gender'],drop_first= True)
```
