```python
import pandas as pd

#Uing CN, CH, LC and MT grades, we want to predict the REDACAO grade 
fields = [ 'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 
          'NU_NOTA_MT', 'NU_NOTA_REDACAO']

dataset = pd.read_csv('../input/microdados-enem-2016-comma-corrected//microdados_enem_2016_coma_corrected.csv', encoding='latin-1', 
                 sep=',', usecols=fields)
dataset.head
len(dataset.index)
```

```python
lenBefore = len(dataset.index)
data_clean = dataset.dropna(subset=["NU_NOTA_CN","NU_NOTA_CH","NU_NOTA_LC","NU_NOTA_MT","NU_NOTA_REDACAO"])
lenAfter = len(data_clean.index)
print("Removed {} entries".format(lenBefore-lenAfter))
print("{} entries left".format(lenAfter))
```

```python
# Calcula a nota total
inputCols = ["NU_NOTA_CN","NU_NOTA_CH","NU_NOTA_LC","NU_NOTA_MT"]
X = data_clean[inputCols]
X.head
```

```python
Y = data_clean["NU_NOTA_REDACAO"]
Y.head
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
train_X, val_X, train_Y, val_Y = train_test_split(X, Y, random_state = 0)
#specify the model. 
#For model reproducibility, set a numeric value for random_state when specifying the model
redacao_model = DecisionTreeRegressor(random_state=1)

# Fit the model
redacao_model.fit(train_X, train_Y)
```

```python
from sklearn.metrics import mean_absolute_error
val_predictions = redacao_model.predict(val_X)
print(mean_absolute_error(val_Y, val_predictions))
```
