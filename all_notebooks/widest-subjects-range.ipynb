```python
df = pd.read_csv("/kaggle/input/moscow-schools-winners-of-educational-olympiads/moscow schools - winners.csv")
df.head()
```

```python
df_clean = df.replace({"НИУ ВШЭ. лицей": "лицей НИУ ВШЭ", 
                       'ГБОУ «Лицей «Вторая школа»': 'ГБОУ лицей «Вторая школа»',
                       'ГБОУ школа № 1329': 'ГБОУ лицей № 1329', 
                       'ГБОУ школа № 1535': 'ГБОУ лицей № 1535',
                       'ГБОУ школа № 179': 'ГАОУ ВО МИОО. СОШ № 179'})
```

```python
school_wide_range = df_clean.groupby("ShortName").Subject.nunique().to_frame().sort_values("Subject", ascending=False).head(5)

widest_range = df_clean[df_clean.ShortName == "лицей НИУ ВШЭ"]["Subject"].value_counts().head(5).to_frame().reset_index().rename(columns={"Subject": "count", "index": "Subject"})

pie, ax = plt.subplots(figsize=[10, 6])
labels = widest_range["Subject"]
x = widest_range["count"]
plt.pie(x=x, labels=labels, autopct="%.1f%%", explode=[0.03]*5, startangle=90)
plt.title("Top 5 subjects of school with widest range", fontsize=14)
```
