    Lets see how both the genders perform in test(math, reading, writing)

```python
for items in df.columns[-3:]:
    sns.barplot(x=df['gender'],y=df[items])
    plt.show()
```

Its clear that female preform well as compare to male. Only in math test males were ahead of females

Lets do the same with groups which group perform well in test

```python
for items in df.columns[-3:]:
    sns.barplot(x=df['race'],y=df[items])
    plt.show()
```

Its clear the in all three test group E preform really well as compare to other groups

**Lets calculate the total marks and their percentages**

```python
df['total'] = (df.math + df.reading + df.writing)
df['precent'] = (df['total'] / 300) * 100
df.head()
```

Find out top 10 students who perform well in test

```python
df.sort_values(by='precent',ascending=False).head(10)
```

We can observe from the above dataset that student who take standart lunch and also completed their preparation test ahead of their exam are in top 10(who have good scores)

**Lets break this dataset into passed_ds (means above 60% are passed)**

```python
passed_ds = df[(df.math > 60) & (df.reading > 60) & (df.writing > 60)]
len(passed_ds)
```

lets calculate the all over percentage of test

```python
passedPrecentage = (len(passed_ds) / len(df)) * 100
passedPrecentage
```

This means that around 57.9% of students passed in the test (who got above 60% in the test)

By analysis the dataset:
* Lunch affect the test score
* Students from group E perform well
* Students who complete their preparation exams before the test got good grades
* There were more female who score are high as compare to males
