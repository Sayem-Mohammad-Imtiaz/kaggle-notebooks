# **Standardize the Variables**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(df.drop('Type',axis=1))
```

```python
scaler_feat = scaler.transform(df.drop('Type',axis=1))
df_feat = pd.DataFrame(scaler_feat,columns=df.columns[:-1])
df_feat.head()
```

# **Train and Test Split**

```python
X = df_feat
y = df['Type']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test,pred))
print(classification_report(y_test, pred))
```

# **Finding the correct value of K**

```python
error_rate = []

for i in range(1, 50):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,10))
plt.plot(range(1,50),error_rate,marker='o')
```

K = 1 will give the best result as compare to other values if we increase k value the error rate will also increased.
