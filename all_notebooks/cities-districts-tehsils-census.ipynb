```python
def toInt(row):
    if(isinstance(row.census_1998,str)):
        s_val = row.census_1998.replace("...","0")
        s_val = s_val.replace(",","")
        i_val = int(s_val)
        return i_val
    return row.census_1998
```

```python
data['census_1998'] = data.apply(lambda row:toInt(row),axis=1)
data.census_1998.head()
```

```python
data.groupby("district").sum().sort_values(by="census_1998",ascending=False) # total 136 districts
```

```python
x = data.groupby("district").sum().sort_values(by="census_1998",ascending=False).reset_index().iloc[:10,0].astype(str)
y = data.groupby("district").sum().sort_values(by="census_1998",ascending=False)["census_1998"].iloc[:10].astype(int)
plt.subplots(figsize=(20,5))
plt.bar(x,y,align='edge',width = 0.3)
fig = plt.gcf()
plt.draw()
plt.show()
fig.savefig('top_10_districts_98.png',dpi=100)
```
