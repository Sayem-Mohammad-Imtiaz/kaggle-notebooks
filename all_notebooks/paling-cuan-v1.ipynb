```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# Input data files are available in the read-only "../input/" directory
# For example, running this (by clicking run or pressing Shift+Enter) will list all files under the input directory

import os
table_sum = []

emitens = pd.read_csv("../input/saham-idx/List Emiten/all.csv")
for emiten in emitens.code:
    filename = "../input/saham-idx/Saham/Semua/"+str(emiten)+".csv"
    #close_first = pd.read_csv(filename,nrows=1).close
    df = pd.read_csv(filename)
    df_first = df.head(1).close
    df_last = df.tail(1).close
    growth = int(df_last) / int(df_first) - 1
    table_sum.append ([str(emiten), round(growth*100,2)])

def sortSecond(val): 
    return val[1]  
  
# list1 to demonstrate the use of sorting  
# using using second key  
  
# sorts the array in ascending according to  
# second element 
table_sum.sort(key = sortSecond, reverse = True) 

from tabulate import tabulate
print(tabulate(table_sum, headers=['Saham', 'Gain/Loss']))




# You can write up to 20GB to the current directory (/kaggle/working/) that gets preserved as output when you create a version using "Save & Run All" 
# You can also write temporary files to /kaggle/temp/, but they won't be saved outside of the current session
```
