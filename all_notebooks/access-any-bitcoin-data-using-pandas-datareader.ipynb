```python
currency = "INR" 
crypto  = "BTC"
```

```python
start = dt.datetime(2021,1,2)
```

```python
end = dt.datetime.now()
```

```python
data = web.DataReader(f"{crypto}-{currency}","yahoo",start,end) 
data1 = web.DataReader(f"ETH-{currency}","yahoo",start,end)
```

```python
data1
```

```python
data
```

```python
mpf.plot(data,type="candle",volume=False,style="yahoo")
```

```python
plt.figure(figsize=(12,5),)
plt.plot(data["Close"],color='g',label="BTC")  
plt.plot(data1["Close"],color='r',label="ETH") 
plt.legend(loc="upper left")
```

```python
data.iplot(theme="henanigans",asImage=True)
```
