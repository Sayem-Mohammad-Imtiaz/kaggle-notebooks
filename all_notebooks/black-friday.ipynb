**Finding which city has highest female to male ratio:**



```python
sns.countplot(x = df.City_Category,hue= df.Gender)
plt.show()
```

```python
df1 = df[['City_Category','Gender']]
df1.head()
```

```python
df1 = pd.get_dummies(df1,columns = ['Gender'])
df1.head()
```

```python
df2 = df1.groupby(['City_Category'])['Gender_F'].agg('sum').reset_index(name="female")
df2.head()
```

```python
df3 = df1.groupby(['City_Category'])['Gender_M'].agg('sum').reset_index(name="male")
df3.head()
```

```python
df2['male'] = df3['male'].values
df2.head()
```

```python
df2['f_to_m_ratio'] = df2['female'].values / df2['male'].values
df2.head()
```

```python
max_ratio = df2['f_to_m_ratio'].max()
index_max_ratio = int(df2[df2['f_to_m_ratio']== max_ratio].index[0])
city_at_index = df2.at[index_max_ratio,'City_Category']
print("ratio of female to male is highest in city - " + city_at_index)
```

**concluded that ratio of female to male is highest in city - B**

**Finding percentage share of  sales for each city :**

```python
df3 = df[['City_Category','Purchase']]
df3.head()
```

```python
df4 = df3.groupby(['City_Category'])['Purchase'].agg('sum').reset_index(name="total_Purchase")
df4.head()
```

```python
plt.pie(df4['total_Purchase'],labels=df4['City_Category'],autopct="%1.1f%%")
plt.legend()
plt.title("Percentage share of purchases for each city")
```
