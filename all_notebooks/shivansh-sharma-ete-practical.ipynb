```python
import pandas as pd
df=pd.read_csv("/kaggle/input/indian-food-101/indian_food.csv")
df
```

```python
print("1. Find percentage of vegetarian and non vegetarian. Show with suitable graph also.")
```

```python
count_veg=len(df[df['diet']=='vegetarian'])
count=len(df)
count_nonveg=count-count_veg
veg=(count_veg/count)*100
nonveg=(count_nonveg/count)*100
print("Percentage of vegetarian dishes is ",veg)
print("Percentage of non-vegetarian dishes is ",nonveg)
```

```python
from matplotlib import pyplot as plt
slices=[count_veg,count_nonveg]
act=["Vegetarian","Non-Vegetarian"]
cols=["c","g"]
plt.pie(slices,labels=act,colors=cols,startangle=90,shadow=True,explode=(0,0.1),autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()
```

```python
#scatterplot
import seaborn as sns
data=df

plt.figure(figsize=(20,10))
sns.scatterplot(x = 'prep_time', y = 'cook_time', s = 50, hue = 'diet', data=data)
plt.title('Cooking Time vs Preperation time', fontsize=16)
plt.xlabel('prep_time', fontsize=12)
plt.ylabel('cook_time', fontsize=12)
```
