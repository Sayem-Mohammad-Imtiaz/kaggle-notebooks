## Import Data

Source from: https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/HTTWYL

```python
protest = pd.read_csv("../input/mass-mobilization-protest-data/mmALL_073120_csv.csv")
protest.head()
```

## Drop columns

```python
protest_df = protest.drop(["id", "ccode", "sources", "notes"], axis = 1)
protest_df
```

## Pull out protests happened after 2019

```python
protest_after2019 = protest_df[protest_df["year"] >= 2019]
protest_after2019.head()
```

## Create Timestamp

Date Int + Month Int + Year Int > Time

```python
# Start time
starttimeInt = pd.DataFrame({"year": protest_after2019['startyear'], "month": protest_after2019['startmonth'], "day": protest_after2019['startday'] })
protest_after2019['starttime'] = pd.to_datetime(starttimeInt[['year', 'month', 'day']])

# End time
endtimeInt = pd.DataFrame({"year": protest_after2019['endyear'], "month": protest_after2019['endmonth'], "day": protest_after2019['endday'] })
protest_after2019['endtime'] = pd.to_datetime(endtimeInt[['year', 'month', 'day']])

# Days
protest_after2019['totaldays'] = protest_after2019['endtime'] - protest_after2019['starttime']
protest_after2019
```
