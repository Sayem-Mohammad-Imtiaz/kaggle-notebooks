**- What is the total irradiation per day?**

```python
def irradiation_per_day(df):
    new_df = df.copy()
    new_df['day'] = new_df.DATE_TIME.dt.day
    new_df['month'] = new_df.DATE_TIME.dt.month
    irradiation_data  = new_df.pivot_table('IRRADIATION',index = ['month','day'])
    irradiation_data.plot(kind = 'bar',figsize = (15,5),rot = 45);
    
irradiation_per_day(Weather_Sensor1)
plt.title('Irradiation Per Day For Plant 1 Weather Sensor');
```

```python
irradiation_per_day(Weather_Sensor2)
plt.title('Irradiation Per Day For Plant 2 Weather Sensor');
```
