```python
 from sklearn import preprocessing
encoder = preprocessing.LabelEncoder()
```

```python
enc_columns =[]
for column in gul.columns:
    if(gul[column].dtype == 'object'):
        enc_columns.append(column)
print(enc_columns)
```

```python
for column in enc_columns:
    gul[column] = encoder.fit_transform(gul[column])
```

```python
gul.head()
```

```python
gul.dtypes
```

```python
gul['Construction chances'].unique()
```

```python
x=gul.iloc[:,5:11]
y=gul['Construction chances']
```

```python
#splitting x and y data to test and train
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression 
classifier = LogisticRegression() 
classifier.fit(x_train, y_train)
```

```python
#predicting on x_test
pred_log = classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
#confusion matrix 
cm = confusion_matrix(y_test, pred_log) 
print ("Confusion Matrix : \n", cm)
```

```python
#finding the accuracy score
from sklearn.metrics import accuracy_score
print ("Accuracy : ", accuracy_score(y_test,pred_log))
```

```python
#Create a classification report for the model
from sklearn.metrics import classification_report
print(classification_report(y_test,pred_log))
```

# **KNN**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
for i in range(1,51,2):
    regressor=KNeighborsClassifier(n_neighbors=i)
    regressor.fit(x_train,y_train)
    pred_knn = regressor.predict(x_test)
    print("Accuracy:",accuracy_score(y_test, pred_knn), "for ",i)
```

```python
#fitting the model
from sklearn.tree import DecisionTreeClassifier
```

```python
d_tree = DecisionTreeClassifier()
```

```python
d_tree.fit(x_train,y_train)
```

```python
pred_tree = d_tree.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy : ",accuracy_score(y_test,pred_tree))
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test,pred_tree))
```

```python
print(classification_report(y_test,pred_tree))
```

## **Grid search CV with KNN**

```python
from sklearn.model_selection import GridSearchCV
param={'n_neighbors':[17]}
est=KNeighborsClassifier()

ht=GridSearchCV(estimator=est, param_grid=param, cv=20)
```

```python
ht.fit(x_train,y_train)
```

```python
pred_cv = ht.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy : ",accuracy_score(y_test, pred_cv))
```

```python
print(confusion_matrix(y_test,pred_cv))
```

```python
print(classification_report(y_test,pred_cv))
```

```python
print ("Accuracy of logestic regression: ", accuracy_score(y_test,pred_log))
print ("Accuracy of KNN: ", accuracy_score(y_test,pred_knn))
print ("Accuracy of decision tree: ", accuracy_score(y_test,pred_tree))
print ("Accuracy of grid search cv: ", accuracy_score(y_test,pred_cv))
```
