### Utilize a biblioteca de pré-processamento para normalizar os atributos de entrada (X)

```python
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
X = SS.fit_transform(X)
```

### Separe o dataset em Treino e Teste

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

### Crie o modelo para MLP, treine-o e faça a predição com os dados de teste (y_pred)

```python
from sklearn.linear_model import Perceptron
pc = Perceptron()
modelo = pc.fit(X_train, y_train)
y_pred = modelo.predict(X_test)
```

### Calcule o Score do Modelo
