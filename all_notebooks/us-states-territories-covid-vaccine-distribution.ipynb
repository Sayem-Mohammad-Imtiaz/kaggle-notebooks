# Which USA state distributed the largest number of vaccines last week?

As of June 8, 2021, California distributed the largest number of vaccines last week.

District of Columbia distributed the largest number of vaccines per capita(100) -- for the last 7 days.



```python
# Used SQL in BigQuery
# Performed on June 9, 2021, so 2021-06-08 was last date with data reported 
# Removed United States as location because want individual states & territories 

SELECT
  location, 
  date,
  CAST(total_distributed AS INT64) AS doses_given
FROM
  proven-solstice-310214.us_daily_covid_vaccine_doses_administered.state_vaccines
WHERE
 date = '2021-06-08' AND location != 'United States'
ORDER BY doses_given DESC LIMIT 10
```

```python
# Used SQL in BigQuery to get info for last 7 days on June 9,2021
SELECT
  location, 
  SUM(distributed_per_hundred) AS per_hundred_doses_given
FROM
  proven-solstice-310214.us_daily_covid_vaccine_doses_administered.state_vaccines
WHERE
 date > '2021-06-01' AND location != 'United States'
GROUP BY  location
ORDER BY per_hundred_doses_given DESC LIMIT 10

```
