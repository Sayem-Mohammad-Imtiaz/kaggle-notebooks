In this kernel we are showing the mean rating of chocolate bars grouped by their cocoa percent. The first step is to load the database in a pandas dataframe.

```python
import pandas as pd

flavors_of_cacao = pd.read_csv('../input/flavors_of_cacao.csv')
```

After loading our dataframe, the next step is to select our columns of interest: Cocoa Percent and Rating.

```python
perc_and_rating = flavors_of_cacao[['Cocoa\nPercent', 'Rating']]
perc_and_rating.head()
```

The next step is to group by cocoa percent.

```python
mean_rating_by_perc = perc_and_rating.groupby(['Cocoa\nPercent'])['Rating'].mean()
mean_rating_by_perc.head()
```

Now that we have the mean ratings by cocoa percent, we just need to sort the values so we can see the percents with the best ratings.

```python
sorted_mean_ratings = mean_rating_by_perc.sort_values(ascending=False)
sorted_mean_ratings.head()
```

We can therefore conclude that according to this dataset the 50% chocolate bars have the highest average rating.
