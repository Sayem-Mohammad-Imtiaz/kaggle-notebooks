## Train-Test Splitting

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(data, test_size=0.2, random_state=42)

print(len(train_set), len(test_set))
```

## Correlations

```python
corr_matrix = train_set.corr()
```

```python
corr_matrix
```

```python
x_train_set = train_set.drop("Purchased", axis = 1)
y_train_set = train_set["Purchased"]

x_test_set = test_set.drop("Purchased", axis = 1)
y_test_set = test_set["Purchased"]
```

```python
x_train_set.shape
y_train_set.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
clf = GaussianNB()
```

```python
clf.fit(x_train_set,y_train_set) #fitting the model 
y_pred = clf.predict(x_test_set) #prediction
y_train_pred = clf.predict(x_train_set)
```

## Confusion matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test_set, y_pred)
```

## Precision and Recall

```python
from sklearn.metrics import precision_score, recall_score
precision_score(y_test_set, y_pred)
```

```python
recall_score(y_test_set, y_pred)
```

## F-Score

```python
from sklearn.metrics import f1_score
f1_score(y_test_set, y_pred)
```

## Accuracy

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test_set, y_pred)
```
