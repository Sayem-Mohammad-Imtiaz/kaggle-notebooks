```python
def Series(Island):
    series = ''
    for char in Island:
        if char.isalpha():
            series += char
    return series

def Sisland(Island):
    sisland = ''
    for char in Island:
        if char.isdigit():
            sisland += char
    if sisland == '':
        return 1
    return int(sisland)
```

```python
ev_data['series']=ev_data['Island'].apply(Series)
ev_data['sisland']=ev_data['Island'].apply(Sisland)
```
