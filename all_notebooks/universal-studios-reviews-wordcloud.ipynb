```python
import numpy as np
import pandas as pd
import glob
import codecs
import requests
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
%matplotlib inline
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('../input/reviewuniversalstudio/universal_studio_branches.csv')
df
```

```python
text_array = df['review_text'].values
text_string = "".join(text_array)
```

```python
stopwords = set(STOPWORDS)
stopwords.add("SPAM_SPAM_SPAM")

wordcloud = WordCloud(width = 881,
                      height = 1038,
                      max_words = 150,
                      stopwords = stopwords,
                      #mask=picture_mask,
                      background_color = "black",
                      contour_width=5,
                      contour_color='steelblue',
                      max_font_size = 200,
                      min_word_length= 2,
                      random_state=40
                     ).generate(text_string)

fig = plt.figure(figsize=(9,10))
plt.imshow(wordcloud)
plt.axis('off')
```
