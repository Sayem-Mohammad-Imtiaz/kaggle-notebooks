##### REMOVING PUNCTUATIONS

```python
Redditnews=train.iloc[:,1:2]
Redditnews
```

```python
Redditnews=train.iloc[:,1:2]
Redditnews.replace('[^a-zA-Z]','',regex=True,inplace=True)

#rename the column
list1=[i for i in range(1)]
new_index=[str(i) for i in list1]
Redditnews.columns=new_index
Redditnews.head()
```

#### coverting headlines to lower case

```python
for index in new_index:
    Redditnews[index]=Redditnews[index].str.lower()
Redditnews.head(1)
```
