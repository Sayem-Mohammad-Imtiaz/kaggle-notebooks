```python
datafm = pd.read_csv("/kaggle/input/football-manager-2020/datafm20.csv")
```

```python
datafm.info()
```

```python
datafm = datafm.drop(columns=["Unnamed: 0"])
datafm.shape
```

```python
datafm.head(10)
```

```python
datafm.describe()
```

```python
print("%d Club" % datafm["Club"].nunique())
print("%d Division" % datafm["Division"].nunique())
print("%d Nation" % datafm["Nation"].nunique())
```

```python
plyrindiv = datafm["Division"].value_counts().head(11)
df_plyrindiv = plyrindiv[1:].to_frame()
df_plyrindiv.plot.bar(figsize=(15,12), fontsize = 14, legend=False)
plt.title("Number of Player (Division)", fontsize=20)
plt.grid(True)
plt.show()
```

```python
plyrinnat = datafm["Nation"].value_counts().head(11)
df_plyrinnat = plyrinnat[1:].to_frame()
df_plyrinnat.plot.bar(figsize=(15,12), fontsize = 14, legend=False)
plt.title("Number of Player (Nation)", fontsize=20)
plt.grid(True)
plt.show()
```

```python
print(datafm["Age"].mean())
print(datafm["Age"].std())
```

```python
plt.figure(figsize=(15,12))
plt.hist(datafm["Age"], bins=37)
plt.title("Players Age Distribution")
plt.grid(linestyle="--")
plt.text(45, 1000, "Mean=24,04; Std=5,5", horizontalalignment='center', verticalalignment='center', fontsize=20)
plt.show()
```
