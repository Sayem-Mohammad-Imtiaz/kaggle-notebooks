```python
def convert_dayTOmonth(days):
    day = int(days) / 365
    convert = '{:5.1f}'.format(day)
    return convert
#just for check the function:
#convert_dayTOmonth(25152)
```

```python
result['DAYS_BIRTH'] = result['DAYS_BIRTH'].apply(convert_dayTOmonth)
```

```python
result
```

```python
result['DAYS_EMPLOYED'] = result['DAYS_EMPLOYED'].apply(convert_dayTOmonth)
```

```python
result.rename(columns={'DAYS_BIRTH':'Age', 'DAYS_EMPLOYED': 'Experience'}, inplace=True)
```

```python
result['AMT_INCOME_TOTAL'] = result['AMT_INCOME_TOTAL'].astype('int')
```

## Most 15 AMT_INCOME_TOTAL:

```python
most = result['AMT_INCOME_TOTAL'].value_counts()[:15]
```

```python
most.index.values
```

```python
def most_income(x):
    if x in most.index.values:
        return x
    else:
        return np.nan
```

```python
result['AMT_INCOME_TOTAL'] = result['AMT_INCOME_TOTAL'].apply(most_income)
```
