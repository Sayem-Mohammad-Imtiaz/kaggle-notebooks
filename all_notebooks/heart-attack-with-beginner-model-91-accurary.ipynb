```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
heart =pd.read_csv('../input/heart-attack-analysis-prediction-dataset/heart.csv')
oxy = pd.read_csv('../input/heart-attack-analysis-prediction-dataset/o2Saturation.csv')
print(heart.info())
print(oxy.info())
```

I random feature to fill in.If accurary lower I delete it.But if it higher I add it.

```python

feature = ['age','cp','trtbps','fbs','restecg','thalachh','exng','oldpeak','slp','caa','thall']
sns.lmplot(x='trtbps',y='thalachh',hue='output',data=heart)
```

```python
X=heart[feature].copy()
y=heart.output.copy()
X_train,X_valid,y_train_,y_valid = train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=5)
```

```python
model=RandomForestClassifier(n_estimators=200,random_state=5)
model.fit(X_train,y_train_)
predict = model.predict(X_valid)
model.score(X_valid,y_valid)
```

```python
print('MAE = '+str(mean_absolute_error(y_valid,predict)))
```

```python
from sklearn.metrics import classification_report, plot_confusion_matrix
fig, ax = plt.subplots(figsize = (10, 6), dpi = 70)
plot_confusion_matrix(model, X_valid, y_valid, ax = ax)
```

```python
output = pd.DataFrame({'Actully' : y_valid,'Output': predict})
output.to_csv('submission.csv', index=True)
```
