```python
#importing datasets
df1 = pd.read_csv("/kaggle/input/ipl-auction-and-ipl-dataset/matches.csv")
df_auc = pd.read_excel("../input/ipl-auction-and-ipl-dataset/Auction.xlsx")
df_batsman = pd.read_excel("/kaggle/input/ipl-auction-and-ipl-dataset/Top_100_batsman.xlsx")
df_bowler = pd.read_excel("/kaggle/input/ipl-auction-and-ipl-dataset/Top_100_bowlers.xlsx")
```

# Data Visualization

```python
# How many times matches were completed or not
sns.countplot(df1["result"])
plt.xlabel("Results of Matches", fontsize=10)
plt.ylabel("No. of Matches" ,fontsize=10)
plt.title("Results of all Matches", fontsize=14)
```

```python
# Number of Matches played at each venue
plt.figure(figsize=(12,8))
a1 = dict(df1['venue'].value_counts())
a1_venue = list(a1.keys())
a1_matches = list(a1.values())
sns.barplot(x=a1_venue[:11],y=a1_matches[:11])
plt.xticks(rotation=90)
plt.ylabel("No. of Matches", fontsize=14)
plt.xlabel("Venues", fontsize=14)
plt.title("Matches Played at Each Venue",fontsize=18)
plt.grid(True , linestyle = '--',linewidth = 0.3)
```

Most IPL Matches are played at Eden Gardens, Kolkata.

# Top 6 buys from 2014-2020

```python
for i in range(0,56,8):
    auction = df_auc.iloc[i:i+8,:].nlargest(5,['PRICE(In crore Indian Rupees) '])
    plt.figure(figsize=(10,6))
    plt.yticks(fontsize=15)
    sns.barplot(x = auction.loc[:,'PLAYER'],y= auction.loc[:,'PRICE(In crore Indian Rupees) '])
    plt.xlabel("PLAYER",fontsize = 15)
    plt.ylabel("PRICE(In crore Indian Rupees) ",fontsize = 15)
    plt.xticks(rotation=45,fontsize = 14)
    plt.title(f"Top 5 buys for Year {auction.iloc[2,4]}",fontsize = 20)
    plt.show()
```

## Top 10 Batsman with Most Runs in IPL

```python
plt.figure(figsize=(16,6))
sns.barplot(x=df_batsman["PLAYER"][:10], y=df_batsman["Runs"][:10])
plt.title("Top 10 Batsman with Most Runs", fontsize=15)
plt.xlabel("Players", fontsize=14)
plt.ylabel("Runs",fontsize=14)
plt.grid(True , linestyle = '--',linewidth = 0.3)
```

```python
Strike_rate = df_batsman.sort_values("SR", ascending=False)
Strike_rate.head()
```

## Top 10 Batsman with Highest Strike Rate

```python
plt.figure(figsize=(16,6))
sns.barplot(x="PLAYER"[:10], y="SR"[:10], data=Strike_rate[:10])
plt.title("Top 10 Batsman with Highest Strike Rate", fontsize=15)
plt.xlabel("Players", fontsize=14)
plt.ylabel("Strike Rate",fontsize=14)
plt.grid(True , linestyle = '--',linewidth = 0.3)
plt.show()
```
