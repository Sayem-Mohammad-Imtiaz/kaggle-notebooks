Impressions are not action-based and are merely defined by a user potentially seeing the advertisement, making CPM campaigns ideal for businesses intent on spreading brand awareness.[https://www.bigcommerce.com/ecommerce-answers/impressions-digital-marketing/#:~:text=Definition%3A%20Impressions%20are%20when%20an,intent%20on%20spreading%20brand%20awareness.]**

> The top 10 advertiser of political ads

```python
data[['advertiser','political']].sort_values(by='political',ascending=False)[:50]
```

**Get top 50 political adverisemnents and thier targetting segments**

```python
adtar = data[['advertiser','political','targets']].sort_values(by='political',ascending=False)[:50]
```

# Each advertisement targetting which segment..it can be Age,Region,Gender and Segment(Political Orientation)

```python
adtar['targeting_segment']= [json.loads(x) for x in adtar['targets']]
```

```python
most_targetted_age_segment=[y.get('segment') for x in adtar['targeting_segment'] for y in x if y.get('segment') != None ]
```

# The most common segment that is targetted for the poliical ad segment is Age - "18 / 18 and Older"

```python
most_common = [item for item in collections.Counter(most_targetted_age_segment).most_common()]
print(dict(most_common))
```

```python
import wordcloud

wordcloud = wordcloud.WordCloud(max_font_size=50, max_words=100, background_color="white").generate_from_frequencies(dict(most_common))
fig = plt.figure(figsize=[10,10])
plt.imshow(wordcloud,interpolation="bilinear")
plt.axis('off')
plt.show()
```

# **Top 10 Advertisers(basically the account that posted the ad on facebook) in the dataset(includes both political and non-political ads) ******

```python
advertisers = data['advertiser'].value_counts()[:10]
plt.figure(figsize=(10,5))
sns.barplot(x=advertisers.index.str.replace(' ', '\n'),y=advertisers.values,alpha=0.8)
plt.title(' Top 10 Advertisers on Facebook')
plt.ylabel('Number of times an account posted the Ad', fontsize=10)
plt.xlabel('advertisers', fontsize=10)
plt.xticks(rotation=308)
plt.show()
```
