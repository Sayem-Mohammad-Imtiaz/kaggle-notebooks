Sometimes simpler is better. This is a small dataset which is almost certainly highly biased to the environment it was sampled from. In a production setting, this would not generalize well. However, creating a placement prediction model doesn't require any special work - we get a ~90% accuracy rate on a test sample of half the data-set using a Random Forest Classifier, after removing columns that are dependent on placement.

```python
data = pd.read_csv("/kaggle/input/factors-affecting-campus-placement/Placement_Data_Full_Class.csv")
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split

y = data["status"]
x = data[["gender", "ssc_p", "ssc_b", "hsc_p", "hsc_b", "degree_p", "workex", "etest_p", "specialisation", "mba_p"]]

xd = pd.get_dummies(x)

trainX, valX, trainY, valY = train_test_split(xd, y, test_size=0.5, random_state=3)
```

```python
xd.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=3, n_estimators=100)
model.fit(trainX, trainY)
```

```python
from sklearn import metrics

predY = model.predict(valX)
accuracy = metrics.accuracy_score(valY, predY)
print("Accuracy: ", accuracy)
```
