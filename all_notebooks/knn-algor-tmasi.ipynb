# KNN Algorithm

The K-NN (K-Nearest Neighbor) algorithm is one of the simplest and most widely used classification algorithms. K-NN is a non-parametric, lazy learning algorithm. If we try to understand the concept of lazy, unlike eager learning, lazy learning does not have a training phase. It does not learn the training data, instead it “memorizes” the training data set. When we want to make an estimate, it looks for the nearest neighbors in the whole dataset.

```python
data["class"]=[1 if i=="Normal"else 0 for i in data["class"]]
```

```python
data.head()
```

```python
y=data["class"].values
x_train=data.drop(["class"],axis=1)
```

# Normalization

Normalization is very important. It is applied so that large numbers in the data do not destroy small numbers during drawing.

```python
#Normalization
x=(x_train-np.min(x_train))/(np.max(x_train)-np.min(x_train))
```

```python
#Train test split 
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state=1)
```

# KNN ALGORİTHM


```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1) #k=1
knn.fit(x_train,y_train)
#Predict
prediction=knn.predict(x_test)

print(" {} nn score: {} ".format(1,knn.score(x_test,y_test)))
```

# KNN Model

```python
score_list=[]
for i in range(1,20):
    knn2=KNeighborsClassifier(n_neighbors=i)
    knn2.fit(x_train,y_train)
    score_list.append(knn2.score(x_test,y_test))
    
#Visualization

plt.plot(range(1,20),score_list)
plt.xlabel("k values")
plt.ylabel("accuracy")
plt.show()
```
