## 1. Mean rating 

```python
print('The mean rating of all the products is:')
a=df['Customer_rating'].mean()
round(a)
```

## 2. Mean according to the delivery

```python
index_names= df[df['Reached.on.Time_Y.N']==0].index
df1= df.drop(index_names)
df1
     # Now we only have products that have been reached on time to the customer
```

```python
print('The mean for the product that have reached on time is:')
b=df1['Customer_rating'].mean()
b
```

```python
index_names= df[df['Reached.on.Time_Y.N']==1].index
df2= df.drop(index_names)
df2
     # Now we only have products that dont have been reached on time to the customer
```

```python
print('The mean for the product that dont have reached on time is:')
c=df2['Customer_rating'].mean()
c
```

So there is a slight difference of rating but nothing significant.

## 3. Rating according to the shipment 

```python
# Calculating the mean rating for each shipment
df1= df[df['Mode_of_Shipment']=='Road']
df1
```

```python
print('The mean for the product that have been transported by road is:')
a=df1['Customer_rating'].mean()
a
```

```python
df2= df[df['Mode_of_Shipment']=='Ship']
df2
```

```python
print('The mean for the product that have been transported by ship is:')
b=df2['Customer_rating'].mean()
b
```

```python
df3= df[df['Mode_of_Shipment']=='Flight']
df3
```

```python
print('The mean for the product that have been transported by flight is:')
c=df3['Customer_rating'].mean()
c
```

```python
# Visualization to match the number of products and the mean of the shipment
plt.bar(range(3), 
        [2.99, 2.99,2.98],
        width =0.5,
        color = 'g',
        edgecolor = 'y', 
        linewidth = 2)
plt.xticks(range(3), ['Road', 'Ship', 'Flight'],
        rotation = 45)
plt.title("The mean according to the shipment", size =16, c= 'b')
plt.ylabel("Mean")

plt.text(-0.1,1.5,"1760")
plt.text(0.88,1.5,"7462")
plt.text(1.89,1.5,"1777")
```
