There is two core object in the pandas, Dataframe and Series.
* Dataframe - A DataFrame is a table. It contains an array of individual entries, each of which has a certain value. Each entry corresponds to a row (or record) and a column.

The array must be same length for each column.

```python
data = pd.DataFrame({'age' : [20,30,35], "name" : ['kishor','aleax','parkhe']})
data.head
```

```python

ages = [20,30,40, 60]

names = ['kishor', 'james', 'alex', 'parkhe']
cities = ['ammsterdam', 'pune', 'barline', 'paris']

data = pd.DataFrame({'age' : ages, "name" : names, 'city' : cities})
```

```python
print(data)
```

# Series
A Series, by contrast, is a sequence of data values. If a DataFrame is a table, a Series is a list. And in fact you can create one with nothing more than a list:

```python
seriesData = pd.Series(['20', '30', '40', '60'])
print(seriesData)
```

A Series is, in essence, a single column of a DataFrame. So you can assign column values to the Series the same way as before, using an index parameter. However, a Series does not have a column name, it only has one overall name. 

```python
seriesData = pd.Series(['20', '30', '40', '60'],index=['2015 Avg. Wind Speed', '2016 Avg. Wind Speed', '2017 Avg. Wind Speed' , '2018 Avg. Wind Speed'], name='Wind Speed Paris')
print(seriesData)
```

```python
data = pd.DataFrame({'age' : ages, "name" : names, 'city' : cities},index=['A', 'B','D','D'])
print(data)
```
