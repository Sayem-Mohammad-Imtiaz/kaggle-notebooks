# Data Visualisation

```python
def Chart_Maker(df,X,Y,c = "label"):
    """
        X - value in X-axis
        Y - value in Y-axis
        c - Color scheme
        """
    chart = alt.Chart(df).mark_bar().encode(
                x = X, #axis
                y = Y,
                color = c        
    )
    display(chart)
```

Let's Start with the Author

```python
Chart_Maker(df = df, X = "count()",Y = "author",c = "label")
```

Most authors have only submitted once and tend to change periodically

```python
Chart_Maker(df = df, X = "score",Y = "count()",c = "label")
print(df['score'].std())
```

Most of the values seem to be concentrated between 0 and 10.
A lot of comments seem to be concentrated on 1 upvote.
Also the data seems to have a high standard deviation the posts with high upvotes are probably skewing the data

```python
Chart_Maker(df = df, X = "subreddit",Y = "count()",c = "label")
```

From these 4 subreddits Cryptocurrency seems to have the most posts follwed by eternium. Most posts seem to be in favour of the CryptoCurrency. This will be a problem while making the model. The model will tend to predict a favourable result.  

```python
df['created_time'] = df['created_time'].apply(parser.parse)# to parse the string as datetime column
Chart_Maker(df = df, X = "created_time",Y = "count()",c = "label")
```

the oldest posts seem to start from November of 2020. It starts of as neither or against as the year progresses into 2021 it becomes against between february and mid-march as the year progresses into april and may it becomes increasingly in favor of Etherium
