# # # Deviation from Normality

This notebook help to understand how to identify if a given series or
returns, are normally distributed or not.

Note: The customized functions refered here are in the python file: edhec_risk_kit.py you can check it: https://www.kaggle.com/rogertovalle/edhec-risk-kit

```python
# Import pandas and customized functions for the risk analysis.
import pandas as pd
import edhec_risk_kit as erk
```

```python
# Get the hfi that contains the returns for each portfolio of the hedge fund
hfi = erk.get_hfi_returns()
```

```python
# Create a dataframe that compares media, median, and show if the median is greater than the median for each protfolio
pd.concat([hfi.mean(), hfi.median(), hfi.mean() > hfi.median()], axis="columns")
```

```python
# Calculate the skewness with customized formula, for each portfolio
erk.skewness(hfi).sort_values()
```

```python
# Calculate the skewness using scipy library
import scipy.stats
scipy.stats.skew(hfi)
```

```python
# Create a randomized series of returns, that is normally distributed.
import numpy as np
normal_rets = np. random.normal(0, .15, size=(26300,1))
normal_rets
```

```python
# Calculate skewness and Kurtosis for the randomized normal distribution series created previously
print("Skewness: \n", erk.skewness(normal_rets))
print("Kurtosis: \n", erk.kurtosis(hfi))
```

```python
# Calculate skewness and kurtosis for hfi series with customized funcions for portfolios
print("Skewness: \n", erk.skewness(hfi))
print("Kurtosis: \n", erk.kurtosis(hfi))
```

```python
# Calculate jarque_bera test to the normal distribution returns.
scipy.stats.jarque_bera(normal_rets)
```

```python
# Calculate Jarque-Bera Test for portfolios
scipy.stats.jarque_bera(hfi)
```

```python
# Use aggregate method of the dataframe to apply Jarque-Bera test for each portfolio,
# using a customized function.

hfi.aggregate(erk.is_normal)
```

```python
# Read the ffme series of ffme returns, and calculate: Skewness and Kurtosis
ffme = erk.get_ffme_returns()
print("Skewness: \n", erk.skewness(ffme), "\n")
print("Kurtosis: \n", erk.kurtosis(ffme), "\n")
print("Jarque-Bera Test: \n:", erk.is_normal(ffme), "\n")
```

```python
# Read the ffme series of ffme returns, and calculate Jarque Bera test for each porfolio
print("Jarque-Bera Test: \n:", ffme.aggregate(erk.is_normal))
```
