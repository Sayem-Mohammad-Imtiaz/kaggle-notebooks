Read The Data

```python
pizza = pd.read_csv('/kaggle/input/pizza-sushi-restaurant-around-jakarta-area/Pizza Around Jakarta Area.csv')
pizza.head()
```

```python
pizza.columns = ['rest_name', 'rest_zone','rating','review','rest_address','rest_type','rest_cost','rest_time']
pizza.head()
```

```python
pizza.groupby('rest_zone')['rating'].mean()
```
