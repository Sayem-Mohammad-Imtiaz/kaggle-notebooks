```python
df=df.drop("Title",axis=1)
df=df.drop("Text",axis=1)
df=df.fillna(0)
df.head(50)
```

```python
df.columns=[i for i in range(len(df.columns))]
df.head(50)
```

```python
reason=[i for i in df[1] if i!=0]
reason=list(dict.fromkeys(reason))
reason
```

```python
victim1=victim2=victim3=0 #counting total number of victims per city
for i in range(len(df)) :
    if df.iloc[i, 0]=="Ghaziabad":
        victim1=victim1+df.iloc[i,4]+df.iloc[i,5]+df.iloc[i,6]
    elif df.iloc[i, 0]=="Kanpur":
        victim2=victim2+df.iloc[i,4]+df.iloc[i,5]+df.iloc[i,6]
    elif df.iloc[i, 0]=="Lucknow":
        victim3=victim3+df.iloc[i,4]+df.iloc[i,5]+df.iloc[i,6]
victim=list((victim1,victim2,victim3))
city = list(("Ghaziabad","Kanpur","Lucknow"))
plt.bar(city,victim,color="red")
plt.xlabel("Cities",fontsize=25)
plt.ylabel("Victims",fontsize=25)
plt.title("City(Murder Data)")
plt.show()



```

```python
reason = np.array(df[1])
reason = reason [reason!= 0]
plt.xticks(rotation = 'vertical')
plt.xlabel('Reason of Murder',fontsize='20')
plt.ylabel('No of Cases',fontsize='20')
plt.title('Reasons(Murders)')
plt.hist(reason,color='green',rwidth=0.8)

            
```
