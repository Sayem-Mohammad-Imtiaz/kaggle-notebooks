```python
import numpy as np 
import pandas as pd

#Подготовка данных
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

#Модели классификации
from sklearn.neural_network import MLPClassifier as mlp
```

```python
df = pd.read_csv("../input/red-wine-quality/winequality-red.csv")
df.head()
```

```python
#Будем классифицировать по качеству вина, 
#для простоты разделим на два класса, плохое вино(качество ниже 6.5) и хорошее
df["good"] = 0
df.loc[df["quality"]>=6.5,"good"] = 1
features = df.columns[:-2]
output = df.columns[-1]
print(f"Признаки: {features.values}\n\n Классы: {output}\n")
```

```python
X = df[features].values
y = df[output].values
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)

print(f"Размер выборки для обучения: {X_train.size}, Размер тестовой выборки: {X_test.size}")
```

Многослойный перцептрон

```python
model_mlp = mlp(hidden_layer_sizes=(50,10, ), activation="tanh", solver="lbfgs", random_state=1,max_iter=30000)
model_mlp.fit(X_train, y_train)
print("Score: ",model_mlp.score(X_test, y_test))
```
