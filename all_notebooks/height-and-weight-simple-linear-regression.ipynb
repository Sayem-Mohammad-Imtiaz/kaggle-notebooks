# Let's go ahead and start solving: $$y=mx+b$$

```python
x = df['Height']
y = df['Weight']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)
```

# We can use np.polyfit for simple regression instead of using sklearn, use help(np.polyfit) for more

# Basically this method accepts x,y and degreee of model trying to fit and returns the coefficients (highest order first)

```python
# Returns highest order coef first!
np.polyfit(x_train,y_train,1)
```

```python
def linear_model(future_height):
    return 61.47893258 * future_height +  -39.19981742
```

```python
predictions = x_test.apply(linear_model)
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test, predictions, color='r')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

# Now we created our model, let's use it for predicting some future height data

```python
future_height = 1.67
linear_model(future_height)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(y_test, predictions)
mae
```

```python
rmse = np.sqrt(mean_squared_error(y_test, predictions))
rmse
```
