```python
# Reading Data from kaggle
data = pd.read_csv("/kaggle/input/salary-data-simple-linear-regression/Salary_Data.csv")
# Reading all the rows with null value
data.dropna(inplace=True)

data.head()
```

```python
# Filtering data
X1 = data['YearsExperience'].values.reshape(-1, 1)
y1 = data['Salary'].values.reshape(-1, 1)
# Spliiting available data
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.3)
```

```python
# fitting data to the model
reg_model1 = LinearRegression()
reg_model1.fit(X_train1, y_train1)

# Predicting the salary for the test data
y_predtest = reg_model1.predict(X_test1)
y_predtrain = reg_model1.predict(X_train1)

# Computing the R-square score of the model
score1 = reg_model1.score(X_test1, y_test1)      #score for the test data
score2 = reg_model1.score(X_train1, y_train1)    #score for the train data 
print(score1, score2)
```

```python
# plotting the predictions and the results of test data together using matplotlib.pyplot
_ = plt.plot(X_test1, y_test1, '.', X_test1, y_predtest)
plt.show()
```

```python
# plotting the predictions and the results of test data together using matplotlib.pyplot
_ = plt.plot(X_train1, y_train1, '.', X_train1, y_predtrain)
plt.show()
```
