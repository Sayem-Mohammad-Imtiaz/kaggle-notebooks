```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('../input/electric-production/Electric_Production.csv', index_col='DATE', parse_dates=True)
df.index.freq = 'MS'
df.head()
```

```python
df.plot(figsize=(15,8), lw=2, grid=True, legend=True);
```

# 1. ETS Decomposition

```python
# ETS Decomposition
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(df['Value'], model='mul')
result.plot();
```

# 2. Simple Moving Average and Exponential Weighted Moving Average

```python
df['SMA 12'] = df['Value'].rolling(window=12).mean()
df['EWMA 12'] = df['Value'].ewm(span=12,adjust=False).mean()
df.plot(figsize=(15,8), lw=2, grid=True, legend=True);
```

# 3. Double Exponential Smoothing - Holts Model

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

df['DES 12'] = ExponentialSmoothing(df['Value'], trend='mul', seasonal_periods=12).fit().fittedvalues.shift(-1)
df.plot(figsize=(15,8), lw=2, grid=True, legend=True);
```

# 4. Triple Exponential Smoothing - Holts Winter Model - Seasonal Trend Model

```python
df['TES 12'] = ExponentialSmoothing(df['Value'],trend='mul',seasonal='mul',seasonal_periods=12).fit().fittedvalues
df.head(20)
```

```python
df.plot(figsize=(15,8), lw=2, grid=True, legend=True);
```

**Note - It can be observed that by using 4 basic forecasting models in this time series dataset, The Double Exponential Smoothing Model (Holts Model) is giving the best results.
