<h2 style='background:brown; border:0; color:white'><center>1. WHAT IS SUPPORT VECTOR MACHINE (SVM)?</center><h2>

<font size="3">Support Vector Machines is considered to be a classification approach, it but can be employed in both types of classification and regression problems. It can easily handle multiple continuous and categorical variables.It can solve linear and non-linear problems and work well for many practical problems.</font>

<a id="2"></a>
<h2 style='background:brown; border:0; color:white'><center>2. SVM TERMINOLOGIES</center><h2>


<h2 style='background:black; border:0; color:white'><center>SUPPORT VECTORS</center><h2>


<font size="3">Support vectors are the data points, which are closest to the hyperplane. These points will define the separating line better by calculating margins. These points are more relevant to the construction of the classifier.Basically, support vectors are imaginary or real data points that are considered landmark points to determine the shape and orientation of the margin.</font>

<h2 style='background:black; border:0; color:white'><center>HYPERPLANE</center><h2>


<font size="3">A hyperplane is a decision plane which separates between a set of objects having different class memberships. Hyperplane will be a **point** in case of 1D data, **line** in case of 2D data, **plane** in case of 3D data and so on.</font>

<h2 style='background:black; border:0; color:white'><center>MARGIN</center><h2>
<font size="3">A margin is a gap between the two lines on the closest class points. This is calculated as the perpendicular distance from the line to support vectors or closest points. If the margin is larger in between the classes, then it is considered a good margin, a smaller margin is a bad margin.</font>

<a id="3"></a>
<h2 style='background:brown; border:0; color:white'><center>3. HOW DOES SVM WORK</center><h2>
    
    
<font size="3">In this algorithm, we plot each data item as a point in n-dimensional space (where n is number of features you have) with the value of each feature being the value of a particular coordinate. Then, we perform classification by finding the optimal hyper-plane that differentiate the two classes very well.
According to the SVM algorithm we find the points closest to the line from both the classes.These points are called support vectors. Now, we compute the distance between the line and the support vectors. This distance is called the margin. Our goal is to maximize the margin. The hyperplane for which the margin is maximum is the optimal hyperplane.</font>

<h2 style='background:black; border:0; color:white'><center>LINEAR SEPARABLE DATA</center><h2>

<font size="3">SVM in case of linear-separable data points works as follows:

* Generate hyperplanes which segregates the classes in the best way. Left-hand side figure showing three hyperplanes black, blue and orange. Here, the blue and orange have misclassified the classes, but the black is separating the two classes correctly.

* Select the right hyperplane with the maximum segregation from the either nearest data points as shown in the right-hand side figure.</font>

<h2 style='background:brown; border:0; color:white'><center>4. SVM KERNELS</center><h2>

<font size='3'>The SVM algorithm is implemented in practice using a kernel. A kernel transforms an input data space into the required form. SVM uses a technique called the kernel trick. Here, the kernel takes a low-dimensional input space and transforms it into a higher dimensional space. In other words, you can say that it converts nonseparable problem to separable problems by adding more dimension to it. It is most useful in non-linear separation problem. Kernel trick helps you to build a more accurate classifier.</font>

<font size='3'> A polynomial kernel is a more generalized form of the linear kernel. The polynomial kernel can distinguish curved or nonlinear input space.It is popular in image processing.Think of the polynomial kernel as a transformer/processor to generate new features by applying the polynomial combination of all the existing features.
    
To illustrate the benefit of applying a polynomial transformer, let’s use a simple example:</font>

<h2 style='background:black; border:0; color:white'><center> RADIAL BASIS FUNCTION(RBF) KERNEL</center><h2>
    
<font size='3'>The Radial basis function kernel is a popular kernel function commonly used in support vector machine classification. RBF can map an input space in infinite dimensional space.It is general-purpose kernel; used when there is no prior knowledge about the data.
    
To illustrate the benefit of applying a Gaussian rbf (gamma = 0.1), let’s use the same example:</font>

Again, it’s impossible for us to find a line to separate the dots (on left hand).
    
But, if we apply Gaussian RBF transformation using two centers (-1,0) and (2,0) to get new features, we will then be able to draw a line to separate the yellow purple dots (on the right):
    
New Feature 1: X_new1 = array([1.01, 1.00, 1.01, 1.04, 1.09])
    
New Feature 2: X_new2 = array([1.09, 1.04, 1.01, 1.00, 1.01])</font>
    

    
<font size='3'>Here gamma is a parameter, which ranges from 0 to 1. 
    
A higher value of gamma will perfectly fit the training dataset, which causes over-fitting. 

Gamma=0.1 is considered to be a good default value. The value of gamma needs to be manually specified in the learning algorithm.</font>

<h2 style='background:BROWN; border:0; color:white'><center>6. TUNING PARAMETERS OF SVM</center><h2>
    
<font size="3">There are some parameters in SVM which can be tuned to bring out the best results from the algorithm.</font>
    


<h2 style='background:black; border:0; color:white'><center>GAMMA</center><h2>    

    
<font size="3">The gamma parameter defines how far the influence of a single training example reaches, with low values meaning ‘far’ and high values meaning ‘close’. In other words, with low gamma, points far away from plausible seperation line are considered in calculation for the seperation line. Where as high gamma means the points close to plausible line are considered in calculation.</font>
    


    



<h2 style='background:black; border:0; color:white'><center>KERNEL</center><h2>

<font size="3">We have various options available with kernels like ‘linear’,‘rbf’,‘poly’ and others. Here ‘rbf’ and ‘poly’ are used for non-linear hyper-plane.</font>

<h2 style='background:black; border:0; color:white'><center>C</center><h2>      

<font size="3">The Regularization parameter (often termed as Penalty parameter C) tells the SVM optimization how much you want to avoid misclassifying each training example.
For large values of C, the optimization will choose a smaller-margin hyperplane if that hyperplane does a better job of getting all the training points classified correctly.Conversely, a very small value of C will cause the optimizer to look for a larger-margin separating hyperplane, even if that hyperplane misclassifies more points.
    
The images below (same as image 1 and image 2) are example of two different regularization parameter. Top one has some misclassification due to lower regularization value. Higher value leads to results like bottom one.</font>

<h2 style='background:brown; border:0; color:white'><center>7. HOW TO PREPARE DATA FOR SVM?</center><h2>

<font size='3'>1. Numerical Inputs: SVM assumes that your inputs are numeric.If you have categorical inputs, you may need to covert them to binary dummy variables (one variable for each category).</font>

<font size='3'>2. SVM is not scale invariant, so it’s highly recommended to scale your data. For example, scale each attribute on the input vector X to [0,1] or [-1,+1], or standardize it to have mean 0 and variance 1.</font>

<h2 style='background:BROWN; border:0; color:white'><center>8. PROS AND CONS ASSOCIATED WITH SVM</center><h2>


<h2 style='background:black; border:0; color:white'><center>PROS</center><h2>  
<font size='3'>
    
* It works really well with clear margin of separation.
    
* It is effective in high dimensional spaces.
    
* It is effective in cases where number of dimensions is greater than the number of samples.
    
* It uses a subset of training points in the decision function (called support vectors), so it is also memory efficient.
    
* SVM Classifiers offer good accuracy and perform faster prediction compared to Naïve Bayes algorithm. </font>


<h2 style='background:black; border:0; color:white'><center>CONS</center><h2>
<font size='3'>
    
* It doesn’t perform well, when we have large data set because the required training time is higher
    
* It also doesn’t perform very well, when the data set has more noise i.e. target classes are overlapping
    
* SVM doesn’t directly provide probability estimates, these are calculated using an expensive five-fold cross-validation. </font>
