```python
from sklearn.preprocessing import LabelEncoder

# performing label encoding
le = LabelEncoder()
df['Outcome'] = le.fit_transform(df['Outcome'])

# converting y again to a dataframe
#y = pd.DataFrame(y)

# determining the datatype of y
#type(y)

df['Outcome'].value_counts()
```

```python
y = df['Outcome']

y.shape
```

```python
X = df.drop(['Outcome'], axis= 1)
X.shape
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 42)

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)

```

```python
from sklearn.svm import SVC

# creating a model
model = SVC()

# feeding the training data into the model
model.fit(x_train, y_train)

# predicting the test set results
y_pred = model.predict(x_test)
```

```python
from sklearn.model_selection import GridSearchCV

param = {
    'C': [0.01,0.05,0.1,0.5,1.0,1.5,5.0,10.0],
    'kernel':['linear', 'rbf'],
    'gamma' :[0.001,0.001,0.1,1.0,10.0]
}
grid_svc = GridSearchCV(model, param_grid = param, scoring = 'accuracy', cv = 10)
```

```python
grid_svc.fit(x_train, y_train)
```

```python
grid_svc.best_params_
```

```python
model = SVC(C = 0.5, gamma = 0.001, kernel = 'linear')
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))

print("Test accuarcy :", model.score(x_test, y_test))
```
