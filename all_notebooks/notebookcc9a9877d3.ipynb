```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn')
```

```python
crimes = pd.read_csv('../input/Chicago_Crimes_2012_to_2017.csv')
```

```python
crimes.drop(['Unnamed: 0', 'Case Number', 'IUCR', 'X Coordinate', 'Y Coordinate','Updated On','Year', 'FBI Code', 'Beat','Ward','Community Area', 'Location', 'District'], inplace=True, axis=1)
```

```python
crimes.head()
```

```python
crimes.Date = pd.to_datetime(crimes.Date, format='%m/%d/%Y %I:%M:%S %p')
```

```python
crimes.info()
```

```python
crimes['Primary Type'].value_counts()
```

```python
crimes['Description'].value_counts()
```

```python
crimes['Location Description'].value_counts()
```

```python
loc_to_change  = list(crimes['Location Description'].value_counts()[20:].index)
desc_to_change = list(crimes['Description'].value_counts()[20:].index)
type_to_change = list(crimes['Primary Type'].value_counts()[20:].index)

crimes['Location Description'][crimes['Location Description'].isin(loc_to_change)] = 'OTHER'
crimes['Description'][crimes['Description'].isin(desc_to_change)] = 'OTHER'
crimes['Primary Type'][crimes['Primary Type'].isin(type_to_change)] = 'OTHER'
```

```python
crimes['Primary Type']         = pd.Categorical(crimes['Primary Type'])
crimes['Location Description'] = pd.Categorical(crimes['Location Description'])
crimes['Description']          = pd.Categorical(crimes['Description'])
```

```python
crimes.index = pd.DatetimeIndex(crimes.Date)
```

```python
crimes.resample('M').size().plot()
```

```python
crimes.resample('D').size().rolling(365).sum().plot()
```

## Crimes Frequency Patterns

```python
hour_by_week = crimes.pivot_table('ID', index=crimes.index.hour, columns=crimes.index.weekday_name, aggfunc=np.size)
```

```python
days = ['Monday','Tuesday','Wednesday',  'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
hour_by_week = hour_by_week[days] # just resort days to make it easier reading the legend
```

```python
hour_by_week.plot(figsize=(13,5), legend=True, alpha=0.9, xticks=np.arange(24))
```

```python
hour_by_month = crimes.pivot_table('ID', index=crimes.index.hour, columns=crimes.index.month, aggfunc=np.size)
```

```python
hour_by_month.plot(figsize=(13,5), legend=False, alpha=.8, xticks=np.arange(24))
```

```python
hour_by_type = crimes.pivot_table(values='ID', columns='Location Description', index=crimes.index.day, aggfunc=np.size)
```

```python
hour_by_type.plot()
```
