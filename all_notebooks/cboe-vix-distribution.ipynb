```python
import pandas as pd
vix_data = pd.read_csv('/kaggle/input/cboe-volatility-index-vix/VIX.csv');
vix_data['Date'] = pd.to_datetime(vix_data['Date'])
vix_data.head()
```

```python
def get_diffTimeFrameAvg(data, timeframe):
    ohlc_dict = {'Open':'first', 'High':'max', 'Low':'min', 'Close': 'last'}
    result = data.resample(timeframe, on='Date').apply(ohlc_dict)
    result['Avg'] = (result['High'] + result['Low'] + result['Open'] + result['Close'])/4;
    return result
```

```python
def get_diffTimeFrameRange(data, timeframe):
    ohlc_dict = {'Open':'first', 'High':'max', 'Low':'min', 'Close': 'last'}
    if (timeframe =='1D'):
        result = data;
        result =result.set_index('Date');
    else:
        result = data.resample(timeframe, on='Date').apply(ohlc_dict);
    result['Range'] = (result['Close'] - result['Open']);
    return result
```

```python
vix_weekly_roll = get_diffTimeFrameRange(vix_data, '1D');
vix_weekly_roll = vix_weekly_roll['Range'].rolling(7).sum()

#vix weekly range distribution
display(vix_weekly_roll)
display(vix_weekly_roll.idxmin(), vix_weekly_roll[vix_weekly_roll.idxmin()])
display(vix_weekly_roll.idxmax(),vix_weekly_roll[vix_weekly_roll.idxmax()])
```

```python
#Vix weekly rolling range distribution
ax=vix_weekly_roll.hist(figsize=[12,8],bins=120)
text='''#vix weekly movement distribution'''
ax.text(x=5, y=300, s=text, fontsize=15,alpha=0.7, color='#334f8d')
```

```python
vix_weekly_roll_med = get_diffTimeFrameRange(vix_data, '1D');
vix_weekly_roll_med = vix_weekly_roll_med['High'].rolling(7).median()

text='''
#vix high weekly median distribution
'''
display(vix_weekly_roll_med)
display(vix_weekly_roll_med.idxmin(),vix_weekly_roll_med[vix_weekly_roll_med.idxmin()])
display(vix_weekly_roll_med.idxmax(),vix_weekly_roll_med[vix_weekly_roll_med.idxmax()])
ax=vix_weekly_roll_med.hist(figsize=[12,8],bins=120)
ax.text(x=30, y=220, s=text, fontsize=15,alpha=0.7, color='#334f8d')
```

```python
#2008 crash and its surroundings

import matplotlib.pyplot as plt

text='''
#Monthly higest median Nov 2008
#Yearly highest median Jun 2009

#Weekly largest fall Nov 2008
#Yearly largest fall Apr 2010

#Monthly largest rise Oct 2008
'''
vix_2008_crash = vix_data[['High', 'Date']];
vix_2008_crash = vix_2008_crash[vix_2008_crash['Date']>pd.to_datetime('01-01-2007')]
vix_2008_crash = vix_2008_crash[vix_2008_crash['Date']<pd.to_datetime('01-01-2012')]
vix_2008_crash.set_index('Date', inplace=True);
ax = vix_2008_crash.plot.line(figsize=[16,8])
x_text_annotation = pd.to_datetime('07-01-2009')
ax.text(x=x_text_annotation, y=50, s=text, fontsize=15,alpha=0.7, color='#334f8d')
```
