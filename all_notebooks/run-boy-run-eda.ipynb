# As I fan of videogame speedrunning , I had to check the progression made by runners throughout the years

```python
# male record progression
male_df = df[df['gender']=='Male']
# Female record progression
female_df = df[df['gender']=='Female']
```

```python
men_time=[]
women_time=[]
for i in male_df['time']:
    (hr,_min,sec) = str(i).split(":")    
    men_time.append((int(hr)*3600)+(int(_min)*60)+(int(sec)))
for i in female_df['time']:
    (hr,_min,sec) = str(i).split(":")    
    women_time.append((int(hr)*3600)+(int(_min)*60)+(int(sec)))
# Time as seconds
male_df['Sec_time'] = men_time
# Time as seconds
female_df['Sec_time'] = women_time
    
```

```python

px.scatter(male_df,x='year',y='Sec_time',labels=df.winner,title="Men's Record Progression : Time v/s Year", color="country")
```

```python
px.scatter(female_df,x='year',y='Sec_time',labels=df.winner,title="Women's Record Progression : Time v/s Year", color="country")
```

# As we can clearly see , the amount of time required by the winner to complete the marathon have been brought lower and lower through the combined effort of many world-class athletes.
And the finding makes sense as over the years various new techniques have been develeped along with better health management and analytics to help runners reach peak human performance
