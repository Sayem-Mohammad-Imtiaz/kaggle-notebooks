```python
#Pivoting Data
dic={"blood_type":["0","A","B","AB"],"Rh":["+","-","-","-"],"Age":[15,25,35,45],"Gender":["F","M","M","F"]}
df=pd.DataFrame(dic)
df
```

```python
df.pivot(index="Age",columns="blood_type",values="Rh")
```

```python
df1=df.set_index(["Age","Gender"])
df1
```

```python
df1.unstack(level=0)
```

```python
df1.unstack(level=1)
```

```python
df2=df1.swaplevel(0,1)
df2
```

```python
pd.melt(df,id_vars="Age",value_vars=["blood_type","Rh"])
```
