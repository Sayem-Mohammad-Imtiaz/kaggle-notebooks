```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.set_printoptions(threshold=np.inf)

selected_country='South Korea'
df = pd.read_csv('../input/novel-corona-virus-2019-dataset/covid_19_data.csv',header=0)
df = df[df['Country/Region']==selected_country]
df = df.groupby('ObservationDate').sum()

print(df)
```

# Day to day cases

```python
df['daily_confirmed'] = df['Confirmed'].diff()
df['daily_deaths'] = df['Deaths'].diff()
df['daily_recovery'] = df['Recovered'].diff()

df['daily_confirmed'].plot()
df['daily_recovery'].plot()
df['daily_deaths'].plot()
plt.show()
```

# Interactive chart

The first peak of the outbreak in South Korea came in late February. On February 18, the 31st patient in South Korea was diagnosed in Daegu. This patient is a member of the Korean religious organization "Shincheonji". After showing symptoms, she still went to gatherings of Shincheonji at least three times. Among the known personnel alone, at least more than 1,000 people participated everytime, which led to the widespread of COVID-19 in Daegu.

As for the confirmed cases at the end of May, according to the Korean Center for Disease Control and Prevention (KCDC), at least 15 people have confirmed virus cases related to Itaewon Club, Itaewon is popular among Koreans and foreigners in the city community.

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_confirmed_object = go.Scatter(x=df.index,y=df['daily_confirmed'].values,name='Daily confirmed')
daily_deaths_object = go.Scatter(x=df.index,y=df['daily_deaths'].values,name='Daily deaths')

layout_object = go.Layout(title='SouthKorea daily cases 20M51955',xaxis=dict(title='Date'),yaxis=dict(title='Number of people'))
fig = go.Figure(data=[daily_confirmed_object,daily_deaths_object],layout=layout_object)
iplot(fig)
fig.write_html('SouthKorea_daily_cases_20M51955.html')
```

 After the case first broke out in China, South Korea always paid attention to the trend of the epidemic and managed to intervene quickly before it became a crisis.
Only a week after the country’s first identified case in late January, government officials met with representatives of several medical companies. They urged these companies to immediately start developing coronavirus detection kits for mass production and promised to give emergency approval.

Within two weeks, although confirmed cases remained not so much, over 10,000 of test kits were shipped every day in South Korea.
South Korea has far more coronavirus tests than any other country, which allows many people to be quarantined and treated shortly after infection, minimizing the spread. If the test result is positive, the health worker will trace the patient's recent activity and track the contact. Once a new case is found, the Korean’s mobile phone will vibrate and an emergency alert will be issued. By early detection and treatment of infections, and isolation of mild patients to special centers, South Korea managed to leave all hospitals to critical patients. The country’s case fatality rate is just over 1%, one of the lowest in the world.

And also, TV broadcasts, subway station announcements, and smartphone reminders will constantly prompt people to wear masks, instruct people how to stay away from others, and report infection data for the day to get people's daily participation.
