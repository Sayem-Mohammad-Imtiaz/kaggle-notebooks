```python
df = pd.read_csv('/kaggle/input/prague-airbnb-price-predict/prague_airbnb_calendar_listing.csv')
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
# definindo o tamanho do conjunto de teste
test_size = 0.20 

seed = 7

# definindo a métrica de avaliação dos algoritmos
scoring = 'neg_mean_squared_error' # MSE
# scoring = 'r2' # R2

# particionando em conjuntos de treino e teste
X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
```

```python
# configurando a validação cruzada em grupos
groups = df['id']
gkfold = GroupKFold(n_splits=3)
gkfold.get_n_splits(X, y, groups=groups)
```

```python
# configuração no do GroupKFold disponivel na pagina do sklearn
groups = df['id']
group_kfold = GroupKFold(n_splits=3)
group_kfold.get_n_splits(X, y, groups)

print(group_kfold)

for train_index, test_index in group_kfold.split(X, y, groups):
#     print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
#     print(X_train, X_test, y_train, y_test)
```

```python
model = LinearRegression()
results = cross_val_score(model, X_train, Y_train, groups=group_kfold, scoring=scoring)
print(abs(results.mean()))
```
