```python
import sklearn
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
```

```python
df_class = pd.read_csv('../input/zoo-animal-classification/class.csv')
df_zoo = pd.read_csv('../input/zoo-animal-classification/zoo.csv')
```

```python
df_class.head(7)
```

```python
df_zoo.head()
```

```python
df_zoo = sklearn.utils.shuffle(df_zoo)
```

```python
X = df_zoo.drop(columns=['animal_name', 'class_type']).values
X = preprocessing.scale(X)
y = df_zoo['class_type'].values
```

```python
test_size = 50
X_train, y_train = X[:test_size], y[:test_size]
X_test, y_test = X[test_size:], y[test_size:]
```

```python
classifier = RandomForestClassifier()
```

```python
classifier.fit(X_train, y_train)
```

```python
classifier.score(X_test, y_test)
```

```python
for X, y in zip(X_test, y_test):
    print('Model: {}\tActual: {}'.format(classifier.predict([X])[0], y))
```
