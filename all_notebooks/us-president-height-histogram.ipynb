# **This code plots the histogram of heights of US Presidents and calculates various types of simple mathematical solution with the help of data.**

**First we have to import requried libraries for this project**

```python
#import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

**Read the dataset**

```python
#reading a .CSV file
data= pd.read_csv("../input/us-president-height-dataset/president_heights.csv")
print(data.head())
```

**Creating array of the height column of dataset**

```python
#creating an array of data
height= np.array(data["height(cm)"])
print(height)
```

**Printing mathematical functions like mean, median etc.**

```python
#printing required results
print("Mean of heights= ", height.mean())
print("Standard Deviation= ", height.std())
print("Minimum height= ", height.min())
print("Maximum height= ", height.max())
print("Median of heights= ", np.median(height))
print("10th percentile of height= ", np.percentile(height, 10))
```

**Plot the histogram with labels**

```python
#visual representation
plt.hist(height)
plt.title("Distribution of US President height")
plt.xlabel("height(cm)")
plt.ylabel("Number")
plt.show()
```
