# Her eyaletin yoksulluk oranı

```python
yoksulluk_yuzdesi.poverty_rate.replace(["-"],0.0,inplace = True)
yoksulluk_yuzdesi.poverty_rate = yoksulluk_yuzdesi.poverty_rate.astype(float)
alan_list = list(yoksulluk_yuzdesi["Geographic Area"].unique())
alan_oranı = []

for i in alan_list:
    x = yoksulluk_yuzdesi[yoksulluk_yuzdesi['Geographic Area']==i]
    yoksulluk = sum(x.poverty_rate)/len(x)
    alan_oranı.append(yoksulluk)
    
data = pd.DataFrame({'alan_list':alan_list,'alan_oranı':alan_oranı})
new_index = (data['alan_oranı'].sort_values(ascending=False)).index.values
sorted_data = data.reindex(new_index)
#visial
plt.figure(figsize= (15,10))
sns.barplot(x=sorted_data['alan_list'],y=sorted_data['alan_oranı'])
plt.xticks(rotation=90)
plt.xlabel('Eyaletler')
plt.ylabel('Yoksulluk Oranı')
plt.title('Yoksulluk Oranı Verilen Eyaletler')
plt.show()
    
```

# Eyaletlerde 25 yaşından büyük nüfusun lise mezuniyet oranı

```python
lise_25_bitirme.head()
```

```python
lise_25_bitirme.percent_completed_hs.value_counts()
```

```python
lise_25_bitirme.percent_completed_hs.replace(["-"],0.0, inplace=True)
lise_25_bitirme.percent_completed_hs=lise_25_bitirme.percent_completed_hs.astype(float)
eyalet_list = list(lise_25_bitirme["Geographic Area"].unique())
mezun_oran = []

for i in eyalet_list:
    x=lise_25_bitirme[lise_25_bitirme["Geographic Area"]==i]
    lise= sum(x.percent_completed_hs)/len(x)
    mezun_oran.append(lise)
data = pd.DataFrame({'eyalet_list':eyalet_list,"mezun_oran":mezun_oran})
new_index = (data["mezun_oran"].sort_values(ascending=True)).index.values
sorted_data2 = data.reindex(new_index)

plt.figure(figsize=(15,10))
sns.barplot(x=sorted_data2["eyalet_list"],y=sorted_data2["mezun_oran"])
plt.xticks(rotation=90)
plt.xlabel("Eyaletler")
plt.ylabel("Liseyi tamamlayan")
plt.title("25 Yaş Üstü Lise Bitiren Eyalet Nüfusunun Yüzdesi")
plt.show()
```

# Siyah, beyaz, yerli ırklara göre eyalet nüfusunun yüzdesi

```python
yaris_sehri.info()
```

```python
yaris_sehri.replace(["-"],0.0,inplace= True)
yaris_sehri.replace(["(X)"],0.0 ,inplace= True)
yaris_sehri.loc[:,['share_white','share_black','share_native_american','share_asian','share_hispanic']] = yaris_sehri.loc[:,['share_white','share_black','share_native_american','share_asian','share_hispanic']].astype(float)
alan_list = list(yaris_sehri["Geographic area"].unique())
share_white = []
share_black = []
share_native_american = []
share_asian = []
share_hispanic = []
for i in alan_list:
    x = yaris_sehri[yaris_sehri["Geographic area"]==i]
    share_white.append(sum(x.share_white)/len(x))
    share_black.append(sum(x.share_black)/len(x))
    share_native_american.append(sum(x.share_native_american)/len(x))
    share_hispanic.append(sum(x.share_hispanic)/len(x))
    share_asian.append(sum(x.share_asian)/len(x))
f,ax = plt.subplots(figsize=(9,15))
sns.barplot(x=share_white,y=alan_list,color='green',alpha = 0.5,label='White' )
sns.barplot(x=share_black,y=alan_list,color='black',alpha = 0.5,label='African American' )
sns.barplot(x=share_native_american,y=alan_list,color='cyan',alpha = 0.5,label='Native American' )
sns.barplot(x=share_asian,y=alan_list,color='yellow',alpha = 0.5,label='Asian' )
sns.barplot(x=share_hispanic,y=alan_list,color='red',alpha = 0.5,label='Hispanic' )

ax.legend(loc='upper right',frameon = True)     # legendlarin gorunurlugu
ax.set(xlabel='Irkların Yüzdes', ylabel='States',title = "Irklara Göre Eyalet Nüfus Yüzdesi ")
plt.show()
```
