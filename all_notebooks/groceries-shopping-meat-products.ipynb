```python
path = "../input/groceries-dataset/Groceries_dataset.csv"
df = pd.read_csv(path)
```

```python
#import packages
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
#show the head data
print(df.head(5))
print(df.tail(5))
print(df.dtypes)
```

```python
df['Date']=pd.to_datetime(df['Date'])
df.dtypes
```

```python
#sort datasets on date
df = df.sort_values('Date')
df.head(5)
```

```python
#find unique values for groceries
print(df['itemDescription'].unique())

#print the unique value per item
result = df['itemDescription'].value_counts()
print(result)
```

```python
#In this analysis, I focus on Meat Products
#Create list
List = ['beef','pork','chicken','hamburger meat','ham','sausage','turkey','meat','organic sausage', 'frozen chicken']
print(List)
```

```python
#Filter dataset on meat shopping
df_meat = df[df.itemDescription.isin(List)]
print(df_meat.head(5))

label = df_meat['itemDescription'].unique()
print(label)
```

```python
#make a circle diagram
# Creating plot 
plt.figure(figsize=(10,10))
plt.pie(df_meat['itemDescription'].value_counts(), autopct='%1.1f%%')
plt.legend(label, loc = "upper right")
plt.tight_layout()
plt.title("Groceries Shopping: Meat Products")
  
# show plot 
plt.show() 
```

```python
#drop values of redundant column for the analysis and add the column month
df_meat = df_meat.drop(['Member_number'], axis=1)
df_meat['Month'] = ""


df_meat['Month'] = df_meat['Date'].dt.month

df_meat.head(5)
```

```python
#pivot table to count the values
pivot = pd.pivot_table(df_meat, index= 'Month',columns='itemDescription',aggfunc='count')

pivot
```

```python
pivot.plot()

plt.title("Groceries Shopping Meat Each Month")
plt.xlabel("Month")
plt.ylabel("Number of Items Sold")
```

```python
#Meat Shopping per Month for various meat products
Months = ['Jan.','Feb.', 'Mar.', 'Apr.','May','Jun.','Jul.','Aug.','Sep.','Oct.', 'Nov.','Dec.']

t = pivot.reset_index()

beef = t.iloc[0:13,1]
chicken = t.iloc[0:13,2]
ham = t.iloc[0:13,4]
hamburger = t.iloc[0:13,5]
organic = t.iloc[0:13,7]
organic = organic.replace(np.nan,0)
pork = t.iloc[0:13,8]
sausage = t.iloc[0:13,9]
turkey = t.iloc[0:13,10]


```

```python
#display the meat sold by month and by type of meat
#subplotting
fig, ax = plt.subplots(4, 2, figsize=(10,7))
fig.tight_layout(pad=3.0)
ax[0][0].plot(Months,beef)
ax[0][1].plot(Months,chicken)
ax[1][0].plot(Months,ham)
ax[1][1].plot(Months,hamburger)
ax[2][0].plot(Months,organic)
ax[2][1].plot(Months,pork)
ax[3][0].plot(Months,sausage)
ax[3][1].plot(Months,turkey)

ax[0, 0].set_title("Meat Shopping: Beef")
ax[0, 1].set_title("Meat Shopping: Chicken")
ax[1, 0].set_title("Meat Shopping: Ham")
ax[1, 1].set_title("Meat Shopping: Hamburger")
ax[2, 0].set_title("Meat Shopping: Organic Sausage")
ax[2, 1].set_title("Meat Shopping: Pork")
ax[3, 0].set_title("Meat Shopping: Sausage")
ax[3, 1].set_title("Meat Shopping: Turkey")



# show plot 
plt.show() 
```

```python
#Assign seasons to months
Seasons = {'Winter': ['December, January, February'], 'Spring': ['March', 'April', 'May'], 'Summer':['June','July','August'], 'Autumn':['September','October','November']}
print(Seasons)
```

```python
season=[]
for row in df_meat['Month']:
    if row == 1:
        season.append('Winter')
    elif row == 2:
        season.append('Winter')
    elif row == 3:
        season.append('Spring')
    elif row == 4:
        season.append('Spring')
    elif row == 5:
        season.append('Spring')
    elif row == 6:
        season.append('Summer')
    elif row == 7:
        season.append('Summer')
    elif row == 8:
        season.append('Summer')
    elif row == 9:
        season.append('Autumn')
    elif row == 10:
        season.append('Autumn')
    elif row == 11:
        season.append('Autumn')
    elif row == 12:
        season.append('Winter')
    else:
        season.append('Not Defined')
```

```python
#create column season and corresponding value
df_meat['Season'] = ""

df_meat['Season'] = season
df_meat.head(5)
```
