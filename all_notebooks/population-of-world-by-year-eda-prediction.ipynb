We have Seen that Population of World Increased Immensly from 1950 to 2020, but the Growth Percentage every Year Decreased, this may be due to compounding effect, before 1990 the Growth is Higher but after 1990 growth start declining till 2020.

```python
df['Rural'] = df.Population - df.Urban
df['RuralPerc'] = (((df.Population-df.Urban)/df.Population)*100)
df['RuralPerc'] = df.RuralPerc.apply(lambda x: int(x))
```

```python
df.drop(['NetChange','ChangePerc'],axis=1,inplace=True)
```

```python
X = df.drop('Population',axis=1)
Y = df.Population
```

```python
x_train = X.head(60)
x_test = X.tail(10)
y_train = Y.head(60)
y_test = Y.tail(10)
```

```python
Model = LinearRegression()
```

```python
Model.fit(x_train,y_train)
```

```python
Model.score(x_test,y_test)
```

```python
MAE = mean_absolute_error(y_true = y_test,y_pred = Model.predict(x_test))
```

```python
print(f'Mean Absolute Error is {MAE}')
```
