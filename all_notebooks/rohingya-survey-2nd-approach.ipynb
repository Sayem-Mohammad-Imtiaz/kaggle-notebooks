```python

nRowsRead = 1000 # specify 'None' if want to read whole file
df_fin = pd.read_csv('../input/cusersmarildownloadsrohingyacsv/rohingya.csv', delimiter=';', encoding = "ISO-8859-1", nrows = nRowsRead)
df_fin.dataframeName = 'rohingya.csv'
nRow, nCol = df_fin.shape
print(f'There are {nRow} rows and {nCol} columns')
df_fin.head()
```

```python
# For my own sake, I will limit the number of columns to those I know the meaning. 

df = df_fin[['Sex','How satisfied are you with the amount of space allocated to you and your family in the camp?','Do you think there are enough healthcare facilities for Rohingya (like aid station, health centre, hospital)?','If the Myanmar government recognizes the rights of Rohingya in the next two years, will you return to Myanmar?', 'If you could choose any country in the world, other than Myanmar, where would you go?', 'How scared are you to leave for another country?', 'Do you think that repatriation to Myanmar will happen in the next two years?', 'How do you feel about the future?']]
```

```python
df.head()
```

```python
df = df.rename(columns={'Sex':'sex', 'How satisfied are you with the amount of space allocated to you and your family in the camp?': 'space', 'Do you think there are enough healthcare facilities for Rohingya (like aid station, health centre, hospital)?': 'healthcare' , 'If the Myanmar government recognizes the rights of Rohingya in the next two years, will you return to Myanmar?': 'return' , 'If you could choose any country in the world, other than Myanmar, where would you go?': 'country' , 'How scared are you to leave for another country?': 'scared' , 'Do you think that repatriation to Myanmar will happen in the next two years?': 'repatriation', 'How do you feel about the future?': 'future'})
```

```python
def plot_maker(cat_feature):
    plot = alt.Chart(df).mark_bar().encode(
        y = cat_feature,
        x = "count()"
    ).interactive()
    return plot
```
