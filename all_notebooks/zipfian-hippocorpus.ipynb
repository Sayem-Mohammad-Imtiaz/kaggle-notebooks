Zipf's law

Zipf's law  is an empirical law formulated using mathematical statistics that refers to the fact that for many types of data studied in the physical and social sciences, the rank-frequency distribution is an inverse relation. The Zipfian distribution is one of a family of related discrete power law probability distributions. It is related to the zeta distribution, but is not identical.
https://en.wikipedia.org/wiki/Zipf%27s_law

```python
def read_docu(file):
    
    all_words = []
    
    with open(file, "r", encoding = "utf-8") as input_file:
        for line in input_file:
            line = line.lower()
            line = line.strip().split()
            all_words += line
        return(all_words)
```

```python
def word_counter(all_words):
    
    word_count = Counter()
    for word in all_words:
        word_count[word] += 1
    return(word_count.values())
```

Zipf Distribution - The Discrete Pareto Distribution

The Zipf distribution is sometimes called the discrete Pareto distribution because it is analogous to the continuous Pareto distribution in the same way that the discrete uniform distribution is analogous to the continuous uniform distribution. https://en.wikipedia.org/wiki/Zipf%27s_law#:~:text=The%20Zipf%20distribution%20is%20sometimes,to%20the%20continuous%20uniform%20distribution.

![](https://image.slidesharecdn.com/zipfdistribution-150713170132-lva1-app6892/95/zipf-distribution-12-638.jpg?cb=1436807185)slideshare

```python
def draw_zipfian_curve(word_count):
    plt.plot(sorted(word_count, reverse = True), marker = "o")
    plt.xscale("log")
    plt.yscale("log")
    plt.xlabel("log(Rank)")
    plt.ylabel("log(Frequency)")
    plt.show()
```

```python
def zipfian_plot(file):
    word_corpus = read_docu(file)
    counts = word_counter(word_corpus)
    draw_zipfian_curve(counts)
```

```python
# Lets first handle numerical features with nan value
numerical_nan = [feature for feature in df.columns if df[feature].isna().sum()>1 and df[feature].dtypes!='O']
numerical_nan
```

```python
df[numerical_nan].isna().sum()
```

```python
## Replacing the numerical Missing Values

for feature in numerical_nan:
    ## We will replace by using median since there are outliers
    median_value=df[feature].median()
    
    df[feature].fillna(median_value,inplace=True)
    
df[numerical_nan].isnull().sum()
```

```python
# categorical features with missing values
categorical_nan = [feature for feature in df.columns if df[feature].isna().sum()>0 and df[feature].dtypes=='O']
print(categorical_nan)
```

```python
#replacing missing values in categorical features
for feature in categorical_nan:
    df[feature] = df[feature].fillna('None')
```
