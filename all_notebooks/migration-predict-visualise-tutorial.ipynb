**Import libraries**

```python
import pandas as pd
from sklearn.cross_validation import train_test_split
from sklearn import svm 
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import numpy as np
from sklearn.naive_bayes import GaussianNB
```

Lets see how the data looks like:
I would like to pay your attention at the column **"Measure"**,"**Country**" and **"CitizenShip**".If we want to get any prediction result we must convert all these string  values to integer.

```python
data = pd.read_csv('../input/migration_nz.csv')
data.head(10)
```

But first lets look what unique values we have got in column "**Measure"**

```python
data['Measure'].unique()
```

Now we need to assign to each uniqie string value it's unique integer value:
In case **when there are not so many values** it is possible to **use "replace" function**.

```python
data['Measure'].replace("Arrivals",0,inplace=True)
data['Measure'].replace("Departures",1,inplace=True)
data['Measure'].replace("Net",2,inplace=True)
```

Now lets check if everything was assigned properly.

```python
data['Measure'].unique()
```


In this case we have got about 250 unique countries.

```python
data['Country'].unique()
```

Now we need to assign to each uniqie string value it's unique integer value:
**I strongly recommend to use .factorize()**

```python
data['CountryID'] = pd.factorize(data.Country)[0]
data['CitID'] = pd.factorize(data.Citizenship)[0]
```

Lets check if everything is ok.

```python
data['CountryID'].unique()
```

Another problem is that we have got missing values.Lets see how many and where they are exactly.

```python
data.isnull().sum()
```

**Fill missing data with median value**

```python
data["Value"].fillna(data["Value"].median(),inplace=True)
```

Lets check if everything is ok.

```python
data.isnull().sum()
```

```python
data.drop('Country', axis=1, inplace=True)
data.drop('Citizenship', axis=1, inplace=True)
```

Lets prepare our data for training and testing(split) 

```python
from sklearn.cross_validation import train_test_split
X= data[['CountryID','Measure','Year','CitID']].as_matrix()
Y= data['Value'].as_matrix()
X_train, X_test, y_train, y_test = train_test_split(
  X, Y, test_size=0.3, random_state=9)
```

Make some predictions.

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(n_estimators=70,max_features = 3,max_depth=5,n_jobs=-1)
rf.fit(X_train ,y_train)
rf.score(X_test, y_test)
```
