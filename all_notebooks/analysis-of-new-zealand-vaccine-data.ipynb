 **Import data**

```python
df = pd.read_csv('../input/covid-vaccination-stats-in-new-zealand/Dates.csv')
df.head()
```

**View Data Information**

```python
df.info()
df.describe()
```

****Convert Time Data****

```python
df['Date'] = pd.to_datetime(df['Date'],format="%Y-%m-%d",errors = 'coerce')
df.head()
```

**A scatter plot of daily vaccinations**

```python
plt.figure(figsize=(10,10))
plt.scatter(df['Date'],df['First dose administered'],label='First dose administered')
plt.scatter(df['Date'],df['Second dose administered'],label='Second dose administered')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Inoculation Quantity')
plt.title('Daily Vaccinations') 
```

```python
b=[]
c=[]
total_1 = 0
total_2 = 0
for i in df['First dose administered']:
    total_1+= i
    b.append(total_1)
df['Total First dose administered'] = b
for i in df['Second dose administered']:
    total_2+= i
    c.append(total_2)
df['Total First dose administered'] = b
df['Total Second dose administered'] = c
df['Just First dose administered'] = df['Total First dose administered'] - df['Total Second dose administered']
df.head()
```

**New Zealand completed the Vaccinations trend**

```python

plt.figure(figsize=(15,10))
plt.plot(df['Date'],df['Total First dose administered'],label='Total First dose administered')
plt.plot(df['Date'],df['Total Second dose administered'],label='Total Second dose administered')
plt.plot(df['Date'],df['Just First dose administered'],label='Just First dose administered')
y_major_locator=MultipleLocator(100000)
ax=plt.gca()
ax.yaxis.set_major_locator(y_major_locator)
plt.ylim(0,1600000)
plt.legend()
plt.title('Total Vaccinations') 
```

```python
df["day"] = df["Date"].dt.day
df["month"] =df["Date"].dt.month
df["year"] = df["Date"].dt.year
```

**Percentage of vaccines completed in New Zealand**

```python
arr = []
sum1 = df['First dose administered'].sum()
sum2 = df['Second dose administered'].sum()
sum3 = sum1-sum2
total_p =5084300
others = 5084300-sum1
arr.append(others)
arr.append(sum3)
arr.append(sum2)

cmap = plt.get_cmap('Greys')
colors = list(cmap(np.linspace(0.45, 0.85, len(arr))))
# Swap in a bright blue for the Lacrosse color.
colors[2] = 'dodgerblue'
labels=['No Vaccine','Just First dose administered','Second dose administered']
labeldistance=1.2
explode=[0,0.1,0.1]
autopct='%3.2f%%'
startangle = 90
pctdistance = 0.6
textprops = {"fontsize":15}
wedgeprops = {"linewidth": 4, 'width':1, "edgecolor":"k"}
fig, ax = plt.subplots(figsize=(10, 12))
plt.pie(arr,
        labels=labels,
        labeldistance=labeldistance, 
        explode=explode,autopct=autopct, 
        startangle = startangle, 
        pctdistance = pctdistance,
        textprops =textprops,
        colors=colors)
ax.set_title('Percentage of vaccines completed in New Zealand', fontsize=18)
plt.axis('equal')
plt.legend()
```
