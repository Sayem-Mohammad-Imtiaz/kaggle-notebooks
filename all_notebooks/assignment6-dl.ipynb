```python
#deleting the column which is not required
r=['PassengerId','Name','Ticket','Cabin']
df1=df.drop(r,axis=1).copy()
df.head()
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
#filling missing Embarked values with the most common values
df1["Embarked"]=df1["Embarked"].fillna(df1["Embarked"].mode()[0])
```

```python
#filling missing age values with the most common values
df1["Age"]=df1["Age"].fillna(df1["Age"].mean())
```

```python
df1.head()
```

```python
cc=['Sex','Embarked']
```

```python
dummy=pd.get_dummies(df1[cc])
dummy.head()
```

```python
df1.head()
```

```python
#remove cc and replace with encoded data
df1.drop(cc,axis=1)
```

```python
df1.head()
```

```python
df1=df1.drop(cc,axis=1)
```

```python
df2=pd.concat([df1,dummy],axis=1)
df2.head()
```

```python
#divide into features and target
X=df2.iloc[:,1:]
y=df2.iloc[:,0]
```

```python
X.head()
```

```python
y.head()
```

```python

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.25)
```

```python
from sklearn import tree
#building decision tree model
dtc=tree.DecisionTreeClassifier(random_state=0)
dtc.fit(X_train, y_train)
```

```python
dtc.score(X_test,y_test)
```

```python
from sklearn.model_selection import train_test_split
Age=[]
Sex=[]
for i in range(1,100):
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.25)
    dtc=tree.DecisionTreeClassifier(random_state=0)
    dtc.fit(X_train, y_train)
    Age.append(i)
    Sex.append(dtc.score(X_test,y_test))

    result=pd.DataFrame()
    result["Age"]=Age
    result["Sex"]=Sex
```

```python
result.describe()
```

```python
result.to_csv("r1.csv")
```
