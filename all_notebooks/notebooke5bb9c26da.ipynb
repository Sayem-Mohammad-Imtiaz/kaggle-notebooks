# Lets now check the previous year data of India and if it has improved over years or not

# We will also compare its data with the fellow competetor China

```python
previous_year_data = pd.read_csv("../input/world-happiness-report-2021/world-happiness-report.csv")
previous_year_data_india = previous_year_data.loc[previous_year_data['Country name']=='India']
previous_year_data_india
```

```python
previous_year_data_china = previous_year_data.loc[previous_year_data['Country name']=='China']
previous_year_data_china
```

```python
fig,(india,china)=plt.subplots(2,1)
india.plot(previous_year_data_india['year'],previous_year_data_india['Life Ladder'])
china.plot(previous_year_data_china['year'],previous_year_data_china['Life Ladder'])
```

# As we can see the life ladder has decreased over time from 2006 to 2019 and again increased in 2020

# But at the same time china which has improved constantly throughout the year

```python
fig,(india,china)=plt.subplots(2,1)
china.plot(previous_year_data_china['year'],previous_year_data_china['Freedom to make life choices'])
india.plot(previous_year_data_india['year'],previous_year_data_india['Freedom to make life choices'])
```

# We had a constant increase in gdp but a little decrease around 2020 but surprisingly China able to maintain the rate of its gdp growth

```python
fig,(india,china)=plt.subplots(2,1)
plt.subplots_adjust(top=5,bottom=3.2)
china.plot(previous_year_data_china['year'],previous_year_data_china['Log GDP per capita'])
india.plot(previous_year_data_india['year'],previous_year_data_india['Log GDP per capita'])
india.set_ylim([0,10])
china.set_ylim([0,10])
india.set_title('INDIA')
china.set_title('CHINA')
```

```python
fig,(india,china)=plt.subplots(2,1)
india.plot(previous_year_data_india['year'],previous_year_data_india['Social support'])
china.plot(previous_year_data_china['year'],previous_year_data_china['Social support'])
india.set_ylim([0,0.9])
china.set_ylim([0,0.9])
india.set_title('INDIA')
china.set_title('CHINA')
plt.subplots_adjust(top=4.5,bottom=3.2)
```
