This is a starter notebook explaining various columns and the values in them.

```python
import pandas as pd
```

This dataset contains only one file: `android-games.csv`.

```python
df = pd.read_csv("../input/top-play-store-games/android-games.csv")
```

```python
df
```

The `rank` columns represents the rank of the game. `title` is the game title that appears on the playstore. `total ratings` represent the total number of ratings that have been cast on the application on play store. The `growth` column represents percentage growth. `price` is the cost of application on play store. `category` is the category of the type as per the play store. The `ratings` represent count of particular ratings on the app store and the last column of `paid` represents if the app is paid or free.

Since `category` is a categorical variable, we will first cast the datatypes

```python
df["category"] = df["category"].astype("category")
```

The list of categories are

```python
df["category"].cat.categories
```

There are no null values but there may be a couple of discrepancies here and there
