### Adding Continents to the table

we have table with the country in it, but what if we want to add the continents in it
we will use python package **"pycountry_convert as pc"**

```python
df['Country']
```

Cleaning the country columns, because many country in columns have abrevated values

```python
index_names = df[df['Country']=='Timor-Leste'].index
df.drop(index_names, inplace=True)
```

```python
df['Country'] = df['Country'].replace(['Hong Kong (S.A.R.)'],'Hong Kong')
df['Country'] = df['Country'].replace(['Libyan Arab Jamahiriya'],'Libya')
df['Country'] = df['Country'].replace(['Venezuela, Bolivarian Republic of...'],'Venezuela')
df['Country'] = df['Country'].replace(['The former Yugoslav Republic of Macedonia'],'North Macedonia')
df['Country'] = df['Country'].replace(['Republic of Korea'],'South Korea')
df['Country'] = df['Country'].replace(['Congo, Republic of the...'],'Congo')
```

```python
pip install pycountry_convert as pc
```

```python
#for adding continents field as according with country

import pycountry_convert as pc

def country_to_continent(country_name):

    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name

# Example
row_count = 0
for i in df['Country']:
  i = str(i)
  if i == 'Other Country (Not Listed Above)' or i == 'nan':
    df.at[row_count,'Continent'] = 'NA'
    continue

  cont = country_to_continent(i)
  df.at[row_count,'Continent'] = cont 
  row_count += 1
```

## Which will be the most desired programming language for the year 2020?

```python
df['LanguageDesireNextYear']
```

```python
#spling the values with their values:

lang_count = {}
for i in df['LanguageDesireNextYear']:
  i  = str(i)
  j = i.split(';')
  for a in j:
    if a in lang_count:
      lang_count[a]+=1
    else:
       lang_count[a] = 1
```

```python
print(lang_count)
```

```python
type (lang_count)
```

```python
# max of a dictionary
lang_count1 = max(lang_count, key = lang_count.get)
print("The most desired languaage for 2020 is: ", lang_count1)
```

## What is the distribution of people who code as a hobby based on gender and continent? **

```python
#categorizing {Men, Women, Other}

Gender = []
for i in df['Gender']:
    if (i=='Man') or (i== 'Man;Non-binary, genderqueer, or gender non-conforming'):
        Gender.append('Man')
    elif (i=='Woman') or (i=='Woman;Non-binary, genderqueer, or gender non-conforming') or (i=='Woman;Man;Non-binary, genderqueer, or gender non-conforming'):
        Gender.append('Woman')
    else:
        Gender.append('Others')
df['Gender'] = Gender
df['Gender'] = df['Gender'].astype('category')
```

```python
df['Gender'].value_counts()
```

```python
Hobbyist = []
for i in df['Hobbyist']:
  if i == 'Yes':
    Hobbyist.append(1)
  else:
    Hobbyist.append(0)
df['Hobbyist'] = Hobbyist
```

```python
df['Hobbyist']
```

```python
df.groupby(['Continent', 'Gender'])['Hobbyist'].count()
```
