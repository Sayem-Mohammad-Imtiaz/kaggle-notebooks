```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
overall_data = pd.read_csv('../input/guns.csv')
suicide_data = overall_data[overall_data.intent == 'Suicide']
homicide_data = overall_data[overall_data.intent == 'Homicide']
```

```python
data_sources = [overall_data, suicide_data, homicide_data]
titles = ['Overall', 'Suicide', 'Homicide']

interested_columns = ['year', 'month', 'intent', 'sex', 'age', 'race', 'place', 'education']
for i, d in enumerate(data_sources):
    for col in interested_columns:
        if col == 'age' or col == 'month':
            d[col].value_counts().sort_index().plot(kind = 'line')
        elif col == 'year' or col == 'education':
            d[col].value_counts().sort_index().plot(kind = 'bar')
        else:
            d[col].value_counts().plot(kind = 'bar')
        plt.title(titles[i] + ': ' + col)
        plt.show()
```

## This was just a quick and dirty analysis of the data
#### Quick Points:
- Overall Gun Violence remained stable throughout these 3 years:
- Gun violence was most prevalant in the summer months
    - This was especially noticable in suicides
- Suicide was the reason behind most deaths by gun
- Men were most likely to die due to gun violence(both for homicides and suicides)
- Most homicide victims were in their 20's; Most Suicide victims were in their 50's
- Most homicide victims were Black; Most Suicide victims were White
- A large majority of victims had a low level of educatiom
    - It was more likely for suicide victims to be moderate to highly educated than homicide victimes
