# Scenario

I am acting as a junior data analyst working for Cyclistic, a bike-share company in Chicago. Therefore, I want to understand how casual riders and annual members use Cyclistic bikes differently. ## The Company

## Process

1. basics
Then I will deal with the data to make it suitable for analysis.
As for the available data, I would like to analyze the difference of bike-type(disability-type if possible), time duration, day of week, time of day, location distribution(may need the Chicago map),and  distance(calculated by lat & lng).

Then I processing these data with mySQL. I'll show the code below, but SQL code are not suitable in this Python environment. I will upload the output of SQL anyway.



```python
# firstly, I create a view to union the 10 sheets, which can simplify my processing.


'''
CREATE VIEW `whole_data_202006-202105` AS
    SELECT *
    FROM `bike-share project`.`202006-divvy-tripdata` 
    UNION SELECT *
    FROM `bike-share project`.`202007-divvy-tripdata` 
    UNION SELECT *
    FROM `bike-share project`.`202008-divvy-tripdata` 
    UNION SELECT *
    FROM `bike-share project`.`202009-divvy-tripdata` 
    UNION SELECT *
    FROM `bike-share project`.`202010-divvy-tripdata` 
    UNION SELECT *
    FROM `bike-share project`.`202011-divvy-tripdata` 
    UNION SELECT *
    FROM `bike-share project`.`202012-divvy-tripdata` 
    UNION SELECT *
    FROM `bike-share project`.`202101-divvy-tripdata` 
    UNION SELECT *
    FROM `bike-share project`.`202102-divvy-tripdata` 
    UNION SELECT *
    FROM `bike-share project`.`202103-divvy-tripdata` 
    UNION SELECT *
    FROM `bike-share project`.`202104-divvy-tripdata` 
    UNION SELECT *
    FROM `bike-share project`.`202105-divvy-tripdata`
'''

# The code are in 3 quotation marks , since it's a Python environment.
# If you want to apply this, remove the quotation marks.
# the organizational structure of the data is as same as the prepare process.
# the output is same as the above string 
```

```python
# process the first table I need

'''
SELECT *
FROM (SELECT
	 ride_id,
     rideable_type,
     CONVERT(started_at, DATE) AS date,
     DAYNAME(started_at) AS day_of_week,
     CONVERT(started_at, TIME) AS start_time,
     timediff(ended_at, started_at) AS time_duration,
     POWER(POWER((end_lat - start_lat) * 111, 2) + POWER((end_lng - start_lng) * 82.64, 2) , 0.5) AS distance,
     member_casual
FROM `whole_data_202006-202105`) AS cool
WHERE
	time_duration > 0 AND distance > 0
'''

# I use subquery cuz mySQL do not support use WHERE to filter the new column I SELECT
# The attribute,distance, calls for some geographical knowledge to get a rough outcome in Chicago
# Anyway, the straight distance is rough for the analysis

# the output is like follow, I upload that to the dataset as well
casual_member_diff = pd.read_csv('../input/d/plumpbear/bikeshare-202006202105/casual_member_diff_processed.csv')
casual_member_diff.head()
```

## Analyze and Visulization

1. trend

formatting the data order by day

calculate the deal numbers each day, and get the trend

```python
'''
SELECT
	c.date,
    casual_count,
    member_count,
    casual_count + member_count AS total_count,
    casual_count / member_count AS casual_to_member
FROM(SELECT
	str_to_date(left(started_at,10),'%Y-%m-%d') AS date,
    count(*) AS casual_count
FROM `whole_data_202006-202105`
WHERE member_casual = 'casual'
GROUP BY date) AS c,
(SELECT
	str_to_date(left(started_at,10),'%Y-%m-%d') AS date,
    count(*) AS member_count
FROM `whole_data_202006-202105`
WHERE member_casual = 'member'
GROUP BY date) AS m
WHERE c.date = m.date
ORDER BY c.date
'''
trend = pd.read_csv('../input/d/plumpbear/bikeshare-202006202105/trend.csv')
trend
```

```python
import plotly as py
import plotly.graph_objs as pygo

pyplt = py.offline.plot

viz_date = trend['date'].tolist() #get the column of date to a list
viz_total_date = trend['total_count'].tolist()
viz_member_date = trend['member_count'].tolist()
viz_casual_date = trend['casual_count'].tolist()

plt.plot(viz_date, viz_total_date, label = 'total')
plt.plot(viz_date, viz_member_date, label = 'member')
plt.plot(viz_date, viz_casual_date, label = 'casual')
plt.legend() # make label applied
plt.title('Trends')
plt.xlabel('time')
plt.ylabel('number of deals')
plt.show()
```

### 2. distribution

The data as above is used to tell the difference of rideable type, weekday, time of the day, duration, and distance.

And then I want to know the difference of member type between stations. So the primary key of this table will be station_id.（start station）

```python
# the 2nd table I need
'''
SELECT
	c.start_station_id,
    casual_count,
    member_count,
    casual_count / member_count AS casual_to_member
FROM(SELECT
	start_station_id,
    count(*) AS casual_count
FROM `whole_data_202006-202105`
WHERE member_casual = 'casual'
GROUP BY start_station_id) AS c,
(SELECT
	start_station_id,
    count(*) AS member_count
FROM `whole_data_202006-202105`
WHERE member_casual = 'member'
GROUP BY start_station_id) AS m
WHERE c.start_station_id = m.start_station_id
ORDER BY c.start_station_id
'''
# I upload it too
# and check that
station_preview = pd.read_csv('../input/d/plumpbear/bikeshare-202006202105/station_diff.csv')
station_preview
```

Then I calculate the whole deal ratio with SQL. 

The code is as follow:

```python
'''
SELECT
	c.casual_count/m.member_count AS deal_ratio
FROM
(SELECT
    count(*) AS casual_count
FROM `whole_data_202006-202105`
WHERE member_casual = 'casual') AS c,
(SELECT
    count(*) AS member_count
FROM `whole_data_202006-202105`
WHERE member_casual = 'member') AS m
'''
```
