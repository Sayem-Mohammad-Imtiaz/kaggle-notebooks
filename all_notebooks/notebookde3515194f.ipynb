```python
train = pd.read_csv('../input/titanic/train.csv')
test = pd.read_csv('../input/titanic/test.csv')
train
```

```python
median_age = train.Age.median()
train.Age = train.Age.fillna(median_age)
train
```

```python
from sklearn.preprocessing import LabelEncoder
le_Sex = LabelEncoder()
```

```python
train["Gender"] = le_Sex.fit_transform(train["Sex"])
train
```

```python
train = train.drop("Sex", axis="columns")
train
```

```python
train = train.drop(["Name", "Ticket", "PassengerId", "Cabin"],axis= "columns")
train
```

```python
train.isnull ().any ()
```

```python
train['Embarked'].mode()[0]
train['Embarked']=train['Embarked'].fillna(train['Embarked'].mode()[0])
train
```

```python
dummies = pd.get_dummies(train.Embarked)
dummies
```

```python
train = pd.concat([train,dummies],axis='columns')
train
```

```python
train = train.drop("Embarked", axis="columns")
train
```

```python
train = train.drop(['PassengerId','Name','Ticket','Cabin'], axis="columns")
train
```

```python
x = train.drop("Survived", axis="columns")
x
```

```python
y = train["Survived"]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20) 
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```
