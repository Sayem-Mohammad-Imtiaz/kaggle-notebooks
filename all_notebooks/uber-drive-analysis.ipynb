Q18. Print all types of purposes for the trip in an array.
Note: This question is based on the dataframe with no 'na' values in the 'PURPOSE' variable.

```python
#Printing all types of purposes for the trip and its miles in an array for the dataset with no missing values 'df'

print(np.array(df['PURPOSE*'].dropna().unique()))
df['MILES*'].groupby(df['PURPOSE*']).sum().sort_values(ascending = False)

# Significant trips are for meetings and customer visit.
```

Q19. Plot a bar graph of Purposes vs Distance.
Note: This question is based on the original dataframe.

```python
# Barplot for Purposes vs Distance for the original dataset "uber_data".

df1 = uber_data.groupby('PURPOSE*')['MILES*'].sum().sort_values(ascending=False).head(10) 
df1
df1= df1.reset_index() # resetting the index
df1
df1.columns = ['PURPOSE*' ,'Total_Miles']
df1
%matplotlib inline
sns.barplot(data= df1 , x= 'PURPOSE*' , y ='Total_Miles').set_title('Purpose vs Distance')
plt.xticks(rotation=90)
```

Q20. Print a dataframe of Purposes and the distance travelled for that particular Purpose.
Note: This question is based on the original dataframe.

```python
# Printing a dataframe for the pupose of the trips and their respective total distance from the original database 'uber_data'.

purpose_dist = uber_data.groupby('PURPOSE*').sum()['MILES*'].sort_values(ascending = False)
print(purpose_dist)
```

Q21. Plot number of trips vs Category of trips. 
Note: This question is based on the original dataframe.

```python
# Plotting number of trips by category from original database 'uber_data'

ax = sns.countplot(x = uber_data['CATEGORY*'], data=uber_data, hue='CATEGORY*', palette="Set3")

# More trips are made for business pupose than the personal purpose.
```

Q22. What is proportion of trips that is Business and what is the proportion of trips that is Personal? 
Note: The proportion calculation is with respect to the 'miles' variable.
Note: This question is based on the original dataframe.

```python
# Proportion of trips between business & personal purpose for the original database 'uber_data'.

categorries_prop = uber_data.groupby('CATEGORY*').agg({'MILES*':'sum'})
categorries_prop
categorries_prop.apply(lambda x: x/x.sum()*100).rename(columns = {'MILES':'% of Miles'})
```
