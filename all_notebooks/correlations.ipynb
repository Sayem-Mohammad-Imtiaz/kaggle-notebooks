```python
import pandas as pd
top50 = pd.read_csv("../input/top50spotify2019/top50.csv", encoding = "latin-1")
```

```python
top50.head()
```

```python
# Correlation matrix
top50[['Beats.Per.Minute', 'Energy', 'Loudness..dB..', 'Speechiness.']].corr()
```

```python
from scipy import stats
pearson_coef, p_value = stats.pearsonr(top50["Energy"], top50["Loudness..dB.."])

# Pearson coefficient / correlation coefficient - how much are the two columns correlated?
print(pearson_coef)

# P-value - how sure are we about this correlation?
print(p_value)
```

# Scatterplot with Regression

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
 
# use the function regplot to make a scatterplot
sns.regplot(x=top50["Energy"], y=top50["Loudness..dB.."],marker="*")
plt.show()
 
# Without regression fit:
# sns.regplot(x=top50["Energy"], y=top50["Loudness..dB.."], fit_reg=False)
# plt.show()
```

```python
slope, intercept, r_value, p_value, std_err = stats.linregress(top50["Energy"],top50["Loudness..dB.."])
print("y = " + str(slope) + "x + " + str(intercept))

# Same as (Pearson) correlation coefficient
print(r_value)
```
