# **График естественного приросты и убыли населения с 1990 по 2017**

```python
df.groupby(['year']).agg({'birth_rate':'mean','death_rate':'mean'}).plot.bar(figsize=(20,10),title='График естественного прироста и убили населения с 1990 по 2017')
plt.show()
```

По графику видно, что тенденция не очень положительная. С 2012 по 2016 года естественный прирост населения в России стал превышать смертность, но в 2017 снова стала прослеживаться отрицательная тенденция. 

# **Топ 10 регионов с самой высокой рождаемостью с 1990 по 2017**

```python
dfTopBirth = df.groupby(['region']).agg({'birth_rate':'mean','death_rate':'mean'}).sort_values(by="birth_rate", ascending=False).head(10)
dfTopBirth
```

```python
dfTopBirth.plot.bar(figsize=(20,10),title='Топ 10 регионов с самой высокой рождаемостью с 1990 по 2017')
plt.show()
```

В тройке лидеров по рождаемости у нас: Чеченская Республика, Республика Тува, Республика Ингушетия. 
И в целом, в топ 10 регионов рождаемость привышает смертность, в период с 1970 по 2017 г. 

# **Топ 10 регионов с самым высоким коэффициентом смертности с 1990 по 2017**
# 

```python
dfTopDeatch = df.groupby(['region']).agg({'birth_rate':'mean','death_rate':'mean'}).sort_values(by="death_rate", ascending=False).head(10)
dfTopDeatch
```

```python
dfTopDeatch.plot.bar(figsize=(20,10),title='Топ 10 регионов с самой высокой смертностью с 1990 по 2017')
plt.show()
```

В тройке лидеров по убыли населения: Псковская область, Тверская область, Новгородская область. Да и в целом, во всей десятке регионов, смертность преобладает более чем в 2 раза, по сравениню с естественным приростом населения, что довольно сильно удручает.

# **Посмотрим на зависимость демографического показателя и количества городского населения**

```python

fig = plt.scatter(df['year'],  df['gdw'], label=None,
            c=df['urbanization'], cmap='viridis',
            linewidth=0, alpha=0.5)
plt.xlabel('Год')
plt.ylabel('Общий демографические показатель')
plt.colorbar(label='% городского населения');
```

В регионах с высоким процентом городского населения общий демографический вес меньше, чем в регионах с небольшим показателем. В больших городах семьи имеют меньше детей, чем в городах и деревнях. Также мы видим, что в первые 10 лет нового века общий демографический вес был значительно ниже, чем в другое время. Это может быть связано с низкой рождаемостью и высокой смертностью в это время.

# **Что там в Орловской области**

```python
df = pd.read_csv('/kaggle/input/russian-demography/russian_demography.csv')
dfOryol = df.loc[df.region == 'Oryol Oblast'].groupby(['year']).agg({'birth_rate':'mean','death_rate':'mean', 'npg':'mean'})
dfOryol
```

```python
dfOryol.plot.bar(figsize=(20,10),title='Демографическая ситуация в Орловской области')
plt.show()
```

В Орловской области, в период с 1990 по 2017 года, каждый год убыль населения преобладает по сравнению с приростом населения.
