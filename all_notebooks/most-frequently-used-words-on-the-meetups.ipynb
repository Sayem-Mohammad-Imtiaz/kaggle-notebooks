```python
font_file_path = '/kaggle/input/lato-font/Lato-Regular.ttf'
mask_file_path = '/kaggle/input/latvia-map/latvia-map.jpg'
text_file_path = '/kaggle/input/rigadsclub/youtube/merged.txt'
```

```python
Image.open(mask_file_path)
```

```python
club_colors = ['#cc1757', '#f8e71c', '#a3ff3e', '#ae72c1']
club_colors = ['#f8e71c', '#a3ff3e']

def random_color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    hsl = Color(np.random.choice(club_colors)).hsl
    h = int(360.0 * hsl[0])
    s = int(100.0 * hsl[1])
    l = int(100.0 * hsl[2])

    return "hsl({}, {}%, {}%)".format(h, s, l)
```

```python
def masked_wordcloud(text, mask, stopwords):
    wordcloud = WordCloud(font_path=font_file_path,
                          background_color='black',
                          stopwords = stopwords,
                          max_words = 15000,
                          max_font_size = 86, 
                          random_state = 42,
                          mask = mask)
    wordcloud.generate(text)
    wordcloud.recolor(color_func = random_color_func)
    figure_size=(40.0,32.0)
    plt.figure(figsize=figure_size)
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
```

# Default stopwords

```python
text=open(text_file_path).read()
mask = np.array(Image.open(mask_file_path))
masked_wordcloud(text, mask, STOPWORDS)
```
