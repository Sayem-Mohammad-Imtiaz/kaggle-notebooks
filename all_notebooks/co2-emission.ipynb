```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
df2 = df.drop(['Make','Model','Vehicle Class','Transmission','Fuel Type'],axis=1)
```

```python
df2.head()
```

```python
X = df2.drop('CO2 Emissions(g/km)',axis=1)
y = df2['CO2 Emissions(g/km)']
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
s = StandardScaler()
```

```python
X_train_s = s.fit_transform(X_train)
X_test_s = s.fit_transform(X_test)
```

```python
lm.fit(X_train_s,y_train)
```

```python
y_pred = lm.predict(X_test_s)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
r2_score(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```
