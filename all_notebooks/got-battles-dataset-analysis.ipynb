 Comparing variables to each other

```python
battles_data.corr()
```

- What does a negative value mean?
An insignificant or non-existing relationship between the variables, their results has nothing to do with each other.

- What does a positive value mean?
Means that there's a dependence within two variables, one may affect the other one with a lower or higher impact. 

- What does a higher number mean versus a lower number?
It will show us if the dependency between variables is bigger or smaller, if its a high number the impact of one variable over the other one will most likely affect the result. Otherwise, it will mean that one variable doesn't necessarily affect the other. This data is more qualitative than quantitative so it will be difficult for us to find interesting correlations within it.

Anyway, in order to understand it better I decide to use a heatmap with the correlation information so it will be easier to identify variables closely related.

```python
#Building the plot
correlation=battles_data.corr()
mp.figure(figsize=(10,10))
sn.heatmap(correlation, vmax=1, square=True,annot=True,cmap='cubehelix')
mp.title('Correlation between each feature',fontsize=20,fontweight="bold")
mp.show()
```

At least know we will know which aspects from the data has higher, lower or none correlation between each other:

- Major capture and the attacker size have a interesting correlation between them. 

- The attacker and defender size has a negative number which means the correlation is insignificant. 

- The battle number and the year has certainly the highest correlation within the chart.
