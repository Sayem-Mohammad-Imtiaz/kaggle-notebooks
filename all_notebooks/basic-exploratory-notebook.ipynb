## States with most number of registered companies

```python
# states_wise_companies = company_data.groupby('REGISTERED_STATE')
fig = plt.gcf()
fig.set_size_inches(22.5, 10.5)
states_wise_companies = company_data.groupby(['REGISTERED_STATE']).COMPANY_NAME.count().sort_values(ascending=False)
states_wise_companies.plot(kind='barh').invert_yaxis()
```

```python
state_wise_growth = company_data[['COMPANY_NAME', 'REGISTERED_STATE', 'DATE_OF_REGISTRATION']]
state_wise_growth.dropna(inplace=True)
state_wise_growth['Time'] = pd.to_datetime(state_wise_growth.DATE_OF_REGISTRATION, errors = 'coerce')
state_wise_growth.drop('DATE_OF_REGISTRATION', axis=1, inplace=True)
```

 # Top 50 oldest companies

```python
top_oldest_companies = state_wise_growth.sort_values('Time')
top_oldest_companies.head(50)
```
