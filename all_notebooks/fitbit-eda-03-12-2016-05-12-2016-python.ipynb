**Question** for analysis of Fitbit data:
- What are some trends in smart device usage?
- How could these trends apply to Bellabeat customers?
- How could these trends help influence Bellabeat marketing strategy?

Bellabeat **products** than can relate to data set:
- Bellabeat app
- Leaf: Bellabeatâ€™s classic wellness tracker

**Lack** of information:
- were users notified that data would be collected befor 03.12.2016 or after, 
- about the technologies used in the data (these are different fitbit models),
- users (gender, age, geographic location, how long the user has been using the device ...),
- calories consumed or burned, logged or calculated,
- units of measure (distance in miles or not),
- how different are VeryActive, FairlyActive, LightlyActive,
- variables were observed, recorded by the user, or calculated (calories, intensity, sleep),
- how the data was collected (data gaps due to non-use of the device or for any other reason),

Data **assumptions** were made:
- Calories, Intensities, MET are calculated based on other metrics 
- Light activity ~ walking, Fair activity ~ walking in a hurry, easy exercises, Very active ~ vigorous exercises, running
- only a fraction of the used devices can measure sleep
- the heart rates of different users (in real numbers) cannot be compared with each other

## Recommendations:

- **divide users of Bellabeat into groups** based on their activity level (remember age, health, etc.) for various measurements of progress, recommendations for improvement, rewards, competitions, etc.
- **prescribe and show levels** for users (within groups) based on the total amount of exercises in the aggregate during the time they were users of Bellabeat products( or loss/gain of weight, resting heart rate etc.)
- encourage users to improve their levels
- conduct a **user survey on the idea of levels**
- remind users to **keep track of their weights** in app 
- **differentiate sleep** (and sleep stages) from sedentary activity 
- users (of different weights, lifestyle, day/night users) need wearable devices of different sizes, comfortable for sleeping, safe around small children, waterproof, etc.
- **Fair Activity**(like walking in a hurry, easy exercises) is not used enough: check activity levels thresholds in app or encourage users in good health to do short bursts of higher level activity
- the category of **LoggedActivitiesDistance** is almost never used, needs to be checked for cancellation
- **the number of users drops** towards the end of the data collection period, check if this is a trend for Bellabeat users, find a period of time when usage is declining, survey users for reasons
- help users know about and **track resting heart rate**
- develop individual metrics for **heart rate levels, show it to user** 
- use heart rate metrics for **measuring stress, sleep and diagnosis of heart problems**
- start **collecting user feedback**, making improvements in these areas, or educating users about the technical limitations of their devices

```python
from scipy.cluster.hierarchy import dendrogram
from sklearn.datasets import load_iris
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import MinMaxScaler


def plot_dendrogram(model, **kwargs):
    # Create linkage matrix and then plot the dendrogram

    # create the counts of samples under each node
    counts = np.zeros(model.children_.shape[0])
    n_samples = len(model.labels_)
    for i, merge in enumerate(model.children_):
        current_count = 0
        for child_idx in merge:
            if child_idx < n_samples:
                current_count += 1  # leaf node
            else:
                current_count += counts[child_idx - n_samples]
        counts[i] = current_count

    linkage_matrix = np.column_stack([model.children_, model.distances_,
                                      counts]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)


scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df_means.values)

# setting distance_threshold=0 ensures we compute the full tree.
model = AgglomerativeClustering(distance_threshold=0, n_clusters=None)

model = model.fit(df_scaled)
plt.title('Hierarchical Clustering Dendrogram')
# plot the top three levels of the dendrogram
plot_dendrogram(model, truncate_mode='level', p=5)
plt.xlabel("Number of points in node (or index of point if no parenthesis).")
plt.show()
```
