**Iris flower dataset with K-Nearest Neighbor algorithm, getting started with basics!**

Here I will find best value for KNN's  **n_neighbor**  by cross validating the **iris flower **dataset.Then,will predict one random feature


**Import libraries:**

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

**Load Dataset:**

```python

Dataset= pd.read_csv("../input/iris-flower-dataset/IRIS.csv")
Dataframe=pd.DataFrame(Dataset) #get dataframe from csv
iris_flower=Dataframe.values #get features and targets

X = iris_flower[:,:-1] #features
y = iris_flower[:,len(iris_flower[0])-1] #targets


```

**Dataset example:**

```python
Dataframe.sample(10)
```

Instantiate **KNeighborsClassifie**r and use **GridSearchCV** for finding best value for n_neighbors by cross validation.
**cv=5**,Dividing the dataset into 5 different sets for test:
 

```python
knn2=KNeighborsClassifier()
param_grid={'n_neighbors':np.arange(1,50)}
knn_gscv=GridSearchCV(knn2,param_grid,cv=5)
knn_gscv.fit(X,y)
k=knn_gscv.best_params_['n_neighbors']
print(knn_gscv.best_params_)
```

**Testing one random array with current n_neightbors=k:**

```python
knn=KNeighborsClassifier(n_neighbors=k)
knn.fit(X,y)
result=knn.predict([[0.1,0.8,0.2,2.1]])
print(result)

```
