```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow
```

```python
train_data = pd.read_csv( "/kaggle/input/loan-data-prediction/Training_Data2021piy.csv")
test_data = pd.read_csv("/kaggle/input/loan-data-prediction/Test_Data2021piy.csv")
```

```python
train_data["profession"]=pd.factorize(train_data.profession)[0]
train_data["city"]=pd.factorize(train_data.city)[0]
train_data["state"]=pd.factorize(train_data.state)[0]
train_data["married"]=pd.factorize(train_data.married)[0]
train_data["house_ownership"]=pd.factorize(train_data.house_ownership)[0]
train_data["car_ownership"]=pd.factorize(train_data.car_ownership)[0]
```

```python
test_data["profession"]=pd.factorize(test_data.profession)[0]
test_data["city"]=pd.factorize(test_data.city)[0]
test_data["state"]=pd.factorize(test_data.state)[0]
test_data["married"]=pd.factorize(test_data.married)[0]
test_data["house_ownership"]=pd.factorize(test_data.house_ownership)[0]
test_data["car_ownership"]=pd.factorize(test_data.car_ownership)[0]
```

```python
X=train_data.iloc[:,1:12]
Y=train_data.iloc[:,12]
```

```python
x=test_data.iloc[:,1:12]
```

```python
%matplotlib inline
from sklearn import linear_model 
from sklearn.preprocessing import  MinMaxScaler
#from sklearn.preprocessing import  MinMaxScaler
sc= MinMaxScaler()
X= sc.fit_transform(X)
Y= Y.values.reshape(-1,1)
Y=sc.fit_transform(Y)
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
x=sc.fit_transform(x)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.utils import plot_model
from numpy import loadtxt
```

```python
model=Sequential()
model.add(Dense(72,input_dim=11,activation='relu'))
model.add(Dense(36,activation='relu'))
model.add(Dense(18,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,epochs=100,batch_size=100,verbose=0)
```

```python
loss,accuracy=model.evaluate(X_test,Y_test,verbose=0)
print('Model Loss: %.2f,Accuracy: %.2f' %((loss*100),(accuracy*100)))
```

```python
predictions=model.predict_classes(X_test)
m=0
for i in range(len(X_test)):
    if predictions[i]==1:
        m=m+1
    print ('Prediction %d= Expected %d'%(predictions[i],Y_test[i])) 
print(m)  
```

```python
prediction=model.predict_classes(x)
n=0
for i in range(len(x)):
    if prediction[i]==1:
        n=n+1
    print ('Prediction %d '%(prediction[i])) 
print(n)  
```

```python
from sklearn.metrics import roc_auc_score
auc = roc_auc_score(Y_test, predictions)
print('ROC AUC: %f' % auc)
```

```python
from sklearn.metrics import classification_report
classification_report(Y_test, predictions)
```
