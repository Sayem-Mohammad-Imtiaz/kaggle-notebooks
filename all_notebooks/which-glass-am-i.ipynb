```python
df=pd.read_csv('/kaggle/input/glass/glass.csv')
df.head()
```

```python
x=df.drop('Type',axis=1)
y=df['Type']

print("value of x: ", x.head())
print("\n value of y: ", y.head())
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
scores={}
scores_list=[]
for i in range(1,100):
    classifer=RandomForestClassifier(n_estimators=i, random_state=0)
    x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2, random_state=100)
    classifer.fit(x_train,y_train)
    y_predict=classifer.predict(x_test)
    scores[i]=metrics.accuracy_score(y_test,y_predict)
    scores_list.append(metrics.accuracy_score(y_test,y_predict))
    
    
```

```python
import matplotlib.pyplot as plt

plt.plot(range(1,100),scores_list)
plt.xlabel('Value of i')
plt.ylabel('Accuracy')
plt.show()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
predictions=classifer.predict(x_test)
print(confusion_matrix(y_test,predictions))
print("Test score:")
print(classifer.score(x_test, y_test))
```

```python
my_submission = pd.DataFrame({'Actual_value': y_test[0:], 'predicted_values':predictions})

my_submission.to_csv('submission.csv', index=False)
```
