```python
train_df = pd.read_csv('../input/titanic_data.csv')
train_df['Title'] = train_df['Name'].str.extract('([A-Za-z]+\.)', expand = True)
train_df['Name'] = train_df['Name'].str.extract('([A-Za-z]+)', expand = True)
train_df['Title'].iloc[796] = train_df['Title'].iloc[796].replace('Dr.','Mrs.')
train_df['Title'] = train_df['Title'].replace(['Countess.','Mlle.','Mme.','Ms.'],'Miss.')
train_df['Title'] = train_df['Title'].replace(['Capt.','Col.','Don.','Jonkheer.','Sir.','Major.','Rev.','Dr.'],'Mr.')
train_df['Title'] = train_df['Title'].replace('Lady.','Mrs.') 

train_df['Family'] = train_df['SibSp'] + train_df['Parch']
g = pd.get_dummies(train_df['Family'],'Fam')
train_df['Alone'] = g.Fam_0
train_df['Embarked'] = train_df['Embarked'].fillna('S')
train_df = train_df.drop(['Cabin','Ticket'], axis = 1)
sex_new ={'male': 1, 'female': 0}
train_df['Sex'] = train_df['Sex'].map(sex_new)
train_df['Age'] = train_df.groupby(['Pclass','Sex','Survived','Title'])['Age'].transform(lambda x: x.fillna(x.mean()))
train_df['Age'] = train_df['Age'].astype(int)
```
