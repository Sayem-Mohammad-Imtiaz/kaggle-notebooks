```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from math import ceil
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

```python
wb_df = pd.read_csv("../input/west-bengal-election-data/west_bengal.csv")
```

```python
wb_df.head()
```

```python
wb_df.education.unique()
```

```python
wb_df["education"] = wb_df.education.replace(['Illiterate','Others','Literate'], 'Illiterate')\
.replace(['Graduate Professional'], 'Graduate').replace(['Post Graduate'], 'Masters')
wb_df["edu"] = wb_df.education.apply(lambda x: {'8th Pass' : 8, '10th Pass' : 10, 'Masters' : 15, 'Graduate Professional': 13, 
 '12th Pass':12, '5th Pass':5, 'Graduate' : 13, 'Illiterate': 0, 'Others' : 0, 'Literate' : 0, 'Doctorate':20}[x])
```

```python
wb_df.head()
```

```python
wb_df.education.value_counts()
```

```python
wb_df.education.value_counts().plot.barh()
```

```python
plt.figure(figsize=[25,8])
res = pd.pivot_table(data=wb_df, index="education", columns="party", values="total_assets", aggfunc=lambda x: round(np.sum(x)/10000000, 3))
sns.heatmap(res, cmap="Greens" , square=True, annot=True)
plt.title("Decleared Asset(1 Core Scale)\n", fontdict={"fontsize":20})
plt.show()
```

```python
plt.figure(figsize=[15,8])
res = pd.pivot_table(data=wb_df, index="education", columns="party", values="criminal_cases", aggfunc=np.sum)
sns.heatmap(res, cmap="Reds" , square=True, annot=True)
plt.title("Decleared Criminal Records\n", fontdict={"fontsize":20})
plt.show()
```
