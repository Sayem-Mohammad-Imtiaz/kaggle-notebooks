```python
#Created a bar graph to compare three different countries in the oecd region

df_gap = gapminder[gapminder.LOCATION == 'LTU']
sns.distplot(df_gap['Value'],  kde=False, label='Lithuania')

df_gap =gapminder[gapminder.LOCATION == 'GRC']
sns.distplot(df_gap['Value'],  kde=False,label='Greece')

df_gap =gapminder[gapminder.LOCATION == 'FRA']
sns.distplot(df_gap['Value'],  kde=False,label='France')

plt.legend(prop={'size': 10})
plt.title('Unemployment in three Countries in the OECD region')
plt.xlabel('Value of Unemployment Raate)')
plt.ylabel('Density')
```
