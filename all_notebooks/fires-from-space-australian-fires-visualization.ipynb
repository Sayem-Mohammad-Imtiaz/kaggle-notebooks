```python
#Importing the following libraries numpy, pandas, os
import numpy as np # https://numpy.org/
import pandas as pd # https://pandas.pydata.org/
import plotly.express as px # https://plotly.com/python/plotly-express/
import os # https://docs.python.org/3/library/os.html

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename)) 
```

```python
# Read in data and view description
df = pd.read_csv("../input/fires-from-space-australia-and-new-zeland/fire_nrt_M6_96619.csv")

# Create a heatmap from the dataset 
fig = px.density_mapbox(df, 
                        lat ='latitude', 
                        lon ='longitude', 
                        z = 'brightness', 
                        color_continuous_scale  = 'inferno',
                        range_color = [200,520],
                        radius = 2,
                        center = dict(lat=-25, lon=130), 
                        zoom = 2.5,
                        mapbox_style = "stamen-toner",
                        animation_frame = "acq_date",
                        )

# Title for visualisation
fig.update_layout(title='Fires in Australia from October 2019 to January 2020')

# Show visualisation
fig.show()
```
