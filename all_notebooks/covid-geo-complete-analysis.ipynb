The process pf cleaning the hashtags column and making a new hashtag column and storing the data in it.
this is done by removing special characters and putting the data int the new column.

```python
def hashtag(x):
  x = str(x)
  x = x.replace('[','')
  x = x.replace(']','')
  x = x.split(',')
  return x
  
df['hashtag'] = df['hashtags'].apply(lambda y : hashtag(y))
df = df.explode('hashtag')
df['hashtag'] = df['hashtag'].astype(str)
df['hashtag'] = df['hashtag'].str.replace("'",'')
df['hashtag'] = df['hashtag'].str.replace('"','')

df['hashtag'] = df['hashtag'].str.replace(" ",'')
df= df[df['hashtag'] != 'nan']
```

```python
plt.figure(figsize=(15,5))
souce = df['hashtag'].value_counts().sort_values(ascending=False)
source = souce.head(10)
source.plot.bar(color=['black', 'red', 'green', 'blue', 'cyan','pink','purple','violet','yellow','orange'])
plt.title('Hashtag  with maximum Usage')
plt.xlabel('Hashtag')
plt.ylabel('Count')
plt.show()
```

Adding time column to the dataset

```python
col_2_list = df['date'].tolist()
col_2_list


lst=[]
for i in col_2_list:
    lst.append(str(i).split(' ')[1])

df['time']= lst
```

```python
df['hashtags'] = df['hashtags'].fillna('[]')
df['hashtag_count'] = df['hashtags'].apply(lambda x : len(x.split(',')))
df.loc[df['hashtags'] == '[]', 'hashtag_count'] = 0
df.head(10)
```

Now let's change months into the name of months i.e Jan,Feb etc

```python
col_3 = df['Dates'].tolist()
lst3 = []
for i in col_3:
  lst3.append(str(i).split('-')[1])
df['month'] = lst3
```

```python
df.head()

def month(x):
  x = str(x)
  x = x.replace('01','Jan')
  x = x.replace('02','Feb')
  x = x.replace('03','March')
  x = x.replace('04','April')
  x = x.replace('05','May')
  x = x.replace('06','June')
  x = x.replace('07','July')
  x = x.replace('08','August')
  x = x.replace('09','Sept')
  x = x.replace('10','Oct')
  x = x.replace('11','Nov')
  x = x.replace('12','Dec')
  return x

df['month'] = df['month'].apply(lambda row : month(row))

```
