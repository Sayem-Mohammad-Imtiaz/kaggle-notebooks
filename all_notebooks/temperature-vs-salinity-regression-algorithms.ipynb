```python
df = pd.read_csv('/kaggle/input/calcofi/bottle.csv')
```

```python
df.head()
```

DATA PREPROCESSING

```python
data = df[['Salnty','T_degC']]
data = data[:][:1000]
data.head()
```

1. REMOVING DUPLICATE VALUES

```python
data.drop_duplicates(subset = ["Salnty", "T_degC"])
```

2. REMOVING NYLL VALUES

```python
data.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy="most_frequent")
imputer.fit(data)
data = imputer.transform(data)
```

3. SCALING THE DATA

```python
from sklearn.preprocessing import StandardScaler
min_max_scaler = StandardScaler()
data= min_max_scaler.fit_transform(data)
data[:5]
```

SELECTING DEPENDENT AND INDEPENDENT VARIABLES

```python
x=data[:,0:1].reshape(-1, 1)
y=data[:,1]
```

PLOTTING THE DATA

```python
import matplotlib.pyplot as mtp 
mtp.scatter(x,y)
mtp.title("Temp vs Salinity")  
mtp.xlabel("Salinity")  
mtp.ylabel("Temp")  
mtp.show()  
```

MAKING TRAIN AND TEST SETS

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state = 42)
```

APPLYING MACHINE LEARNING MODELS

1. LINEAR REGRESSION

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(x_train,y_train)
```

```python
pred = reg.predict(x_test)
```

```python
import matplotlib.pyplot as mtp 
mtp.scatter(x_test,y_test,color="blue")  
mtp.scatter(x_test,pred, color="red")  
mtp.title("TEMPERATURE VS SALINITY(Linear Regression)")  
mtp.xlabel("Salinity")  
mtp.ylabel("Temperature")  
mtp.show()  
```

```python
accuracy = reg.score(x_test, y_test)
print(accuracy)
```

```python
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error
rmse = mean_squared_error(y_test, reg.predict(x_test), squared=False)
print("RMSE: %.2f"% rmse)
mae = mean_absolute_error(y_test, pred)
print('Mean absolute error: %.2f'%mae)
print("R2-score: %.2f" % r2_score(np.asanyarray(y_test), pred))
```

2. POLYNOMIAL REGRESSION

```python
from sklearn.preprocessing import PolynomialFeatures  
poly_regs= PolynomialFeatures(degree= 2)  
x_poly= poly_regs.fit_transform(x_train)  
lin_reg_2 =linear_model.LinearRegression()  
lin_reg_2.fit(x_poly, y_train)  
```

```python
mtp.scatter(x_test,y_test,color="blue")  
mtp.scatter(x_test, lin_reg_2.predict(poly_regs.fit_transform(x_test)), color="red")  
mtp.title("TEMPERATURE VS SALINITY(Polynomial Regression)")  
mtp.xlabel("Salinity")  
mtp.ylabel("Temperature")  
mtp.show()  
```

```python
accuracy = lin_reg_2.score(poly_regs.fit_transform(x_test), y_test)
print(accuracy)
```

```python
pred1 = lin_reg_2.predict(poly_regs.fit_transform(x_test))
```

```python
from sklearn.metrics import mean_squared_error,r2_score, mean_absolute_error
rmse = mean_squared_error(y_test, pred1, squared=False)
print("RMSE: %.2f"%rmse)
mae = mean_absolute_error(y_test, pred1)
print('Mean absolute error: %.2f'%mae)
print("R2-score: %.2f" % r2_score(np.asanyarray(y_test), pred1))
```
