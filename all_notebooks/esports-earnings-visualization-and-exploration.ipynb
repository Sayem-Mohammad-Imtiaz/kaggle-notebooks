<h1 style="text-align:center; background-color: SeaShell; padding: 0.6em;">eSports Earnings: Data Visualization and Exploration</h1>

```python
import pandas as pd 
df = pd.read_csv('../input/esports-earnings-for-players-teams-by-game/highest_earning_teams.csv')
```

```python
df.TotalUSDPrize = df.TotalUSDPrize.round().astype('int')
df.head(3).style.set_properties(**{'background-color': 'LightBlue'}, subset=['TotalUSDPrize']) 
```

```python
HighestTotalUSDPrize = df["TotalUSDPrize"].max();
print("Highest TotalUSDPrize: " , HighestTotalUSDPrize, "$")
```

Highest TotalUSDPrize around 34 million dollars.

```python
print("                   Team with the highest TotalUSDPrize:") 
df[df["TotalUSDPrize"] == HighestTotalUSDPrize].style.set_properties(**{'background-color': '#e89ca5'}, subset=['TotalUSDPrize'])
```

Is this team participated in the highest number of tournaments? No, there another team. That even participated in a much bigger number of tournaments.

```python
print("         Team with the highest TotalTournaments:")
df[df.TotalTournaments == df["TotalTournaments"].max()].style.set_properties(**{'background-color': 'AntiqueWhite'}, subset=['TotalTournaments'])
```

```python
genre_unique = list(df.Genre.unique())
genre_unique.sort()
genre_df = pd.DataFrame(index=range(0,len(genre_unique)), columns = ["Genre", "GenereTotalUSD", "GenereTotalTournaments", "GenereTournamentMean"])

for i, name in enumerate(genre_unique):
    genre_df.iat[i,0] = name
    genre_df.iat[i,1] = df[df.Genre == name].TotalUSDPrize.sum()
    genre_df.iat[i,2] = df[df.Genre == name].TotalTournaments.sum()
    genre_df.iat[i,3] = genre_df.iat[i,1] / genre_df.iat[i,2]
    
genre_df.index = genre_df["Genre"]    
```

```python
pd.set_option('display.float_format', '{:.0f}'.format)
genre_df
```

```python
genre_df.GenereTotalUSD.plot.pie(title="GenereTotalUSD per Genre\n");
```

Much more total revenue in the genre 'Multiplayer Online Battle Arena'.<br><br> Are there a correlation between total amount of prices and popularity of the games? For measure popularity, I collected data about amount of Twitter followers of the games.

```python
games_names = df.Game.unique()
games_names.sort()

games_popularity_df = pd.DataFrame(columns=["GameName", "GameFollowers", "TwitterLink"],index=games_names)
games_popularity_df.loc["Arena of Valor"] = ["Arena of Valor", "63100", "twitter.com/ArenaofValor"]
games_popularity_df.loc["Counter-Strike: Global Offensive"] = ["Counter-Strike: Global Offensive", "1000000", "twitter.com/csgo"]
games_popularity_df.loc["Dota 2"] = ["Dota 2", "997600", "twitter.com/DOTA2"]
games_popularity_df.loc["Fortnite"] = ["Fortnite", "12500000", "twitter.com/FortniteGame"]
games_popularity_df.loc["Hearthstone"] = ["Hearthstone", "825800", "twitter.com/PlayHearthstone"]
games_popularity_df.loc["Heroes of the Storm"] = ["Heroes of the Storm", "448400", "twitter.com/blizzheroes"]
games_popularity_df.loc["League of Legends"] = ["League of Legends", "4800000", "twitter.com/leagueoflegends"]
games_popularity_df.loc["Overwatch"] = ["Overwatch", "3300000", "twitter.com/PlayOverwatch"]
games_popularity_df.loc["PUBG"] = ["PUBG", "1700000", "twitter.com/PUBG"]
games_popularity_df.loc["Starcraft II"] = ["Starcraft II", "391900", "twitter.com/StarCraft"]

for i in games_names:
    game_genre = df[df.Game == i].head(1).Genre.item()
    games_popularity_df.loc[i,"Genre"] = game_genre
    
games_popularity_df.GameFollowers = games_popularity_df.GameFollowers.astype('int64')

for i in games_names:    
    games_popularity_df.loc[i,"GenreFollowers"] = games_popularity_df.GameFollowers[games_popularity_df.Genre == games_popularity_df.loc[i,"Genre"]].sum()

games_popularity_df
```

```python
for i in genre_df.index:
    genre_df.loc[i,"GenreFollowers"] = games_popularity_df[games_popularity_df.Genre == i].head(1).GenreFollowers.item()
```

```python
genre_df.plot.scatter(x="GenereTotalUSD", y="GenreFollowers", s=110, figsize=(14,8),grid=True, title="GenereTotalUSD and GenreFollowers\n");
```

Some positive correlation between GenereTotalUSD and GenreFollowers.
