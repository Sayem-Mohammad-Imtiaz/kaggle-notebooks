```python
df = df.replace('Dadra and Nagar Haveli and Daman and Diu', 'Daman & Diu')
```

```python
px.bar(df, x="State/UTs", y="Total Cases",color="Death Ratio (%)", title="Total Cases as per each State : ",width=1000, height=600)
```

## Observation :
### Hightest Number of Cases : Maharashtra
### Hightes Death Ratio : Punjab

```python
px.bar(df, x="State/UTs", y="Total Cases", color="Deaths",width=1000, height=600)
```

## Observation:
### Highest Death : Maharashtra
### Lowest Death : Andaman And Nicobar

```python
px.scatter_3d(df, x="Active Ratio (%)", y="Discharge Ratio (%)", z="Total Cases", color="State/UTs",size_max=18, opacity=0.7)
```

## Observation :
### Hightest Discharge Ratio : Daman And Diu
### Lowest Discharge Ratio: Mizoram

```python
px.scatter_3d(df, x="Total Cases", y="Active", z="Active Ratio (%)", color="State/UTs")
```

## Observation :
### Highest Active Cases:Maharashtra
### Lowest Active Cases: Andaman And Nicobar

```python
px.scatter(df, x="Total Cases", y="Deaths", size="Death Ratio (%)", hover_name="State/UTs",log_x=True, size_max=30)
```

## Obervation :
### Hightest Death Ratio: Maharashtra 
### Lowest Death Ratio: Andaman And Nicobar

## Discharge Ratio Per State:

```python
px.scatter(df, x="Total Cases", 
           y="Discharge Ratio (%)",
           size="Discharge Ratio (%)", 
           color="Discharge Ratio (%)",
           hover_name="State/UTs", 
           log_x=True,
           size_max=20)
```

## Observation :
### Highest Discharge Ratio : Daman And Diu
### Lowest Discharge Ratio : Mizorum 

```python
num_cols = df.select_dtypes(exclude="object").columns.to_list()
m=0
for i in num_cols:
    
    fig,ax = plt.subplots(1,2)
    
    sns.histplot(data=df, x=i, kde=True, color=sns.color_palette()[m+1], ax=ax[0])
    
    sns.boxplot(data=df, x=i,color=sns.color_palette()[m+1], ax=ax[1])

    m+=1
    
    plt.show()
```

## Observation :
### We have outliers in Total Cases, Active, Discharged, Deaths, Active Ratio (%) and Discharge Ratio.
