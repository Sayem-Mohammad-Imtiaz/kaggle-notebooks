```python
X = train.drop('price_range',axis=1) #dropped the target variable
```

```python
X.shape
```

```python
X.head()
```

```python
Y=train['price_range']
```

```python
Y.head()
```

```python
test=test.drop('id',axis=1)
```

```python
test.shape
```

```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
X_std=std.fit_transform(X)
X_std
```

```python
test_std=std.transform(test)
test_std
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
X_test
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
l=LogisticRegression()
```

```python
X_train_std=std.fit_transform(X_train)
```

```python
X_test_std=std.transform(X_test)
```

```python
l.fit(X_train_std,Y_train)
```

```python
Y_pred=l.predict(X_test_std)
```

```python
Y_pred
```

```python
from sklearn.metrics import accuracy_score
l_accuracy = accuracy_score(Y_test,Y_pred)
```

```python
l_accuracy
```

```python
from sklearn.neighbors import KNeighborsClassifier #knn
knn=KNeighborsClassifier()
knn.fit(X_train_std,Y_train)
```

```python
Y_pred=knn.predict(X_test_std)
```

```python
Y_pred
```

```python
knn_accuracy = accuracy_score(Y_test,Y_pred)
```

```python
knn_accuracy
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,Y_train)
```

```python
Y_pred=dt.predict(X_test)
```

```python
Y_pred
```

```python
dt_accuracy = accuracy_score(Y_test,Y_pred)
```

```python
dt_accuracy
```

```python
sns.barplot(x = [l_accuracy,knn_accuracy,dt_accuracy], y = ['Linear Regression','KNN','Decision Tree'],palette = 'rocket')
```
