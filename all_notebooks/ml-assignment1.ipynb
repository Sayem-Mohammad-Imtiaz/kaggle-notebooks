```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 

#from numpy import pi
#import os
```

```python
d = pd.read_csv("../input/up-crime-data-year-2019/MR data - Compiled Data Set.csv")
d
```

```python
d = d.drop(['Title'], axis=1)
d = d.drop(['Text'], axis=1)
d
```

```python
d = d.fillna(0)
col = d.columns
d.columns = [i for i in range(0,len(d.columns))] 
d
```

```python
d =d.replace(["Property Disputes", "Family Dispute", "Petty Quarrels", "Money Disputes",  "Personal Vendetta",  "Love Affairs",  "Casteism", "Unknown reasons" ],[1,2,3,4,5,6,7,8])
d.head(20)
```

```python
d[11]=0
d[12]=0
for i in range(0,len(d)):
    if ',' in str(d.iloc[i,8]):
        cl=d.iloc[i,8].split(',')
        d.iloc[i,8]=int(cl[0])
        d.iloc[i,11]=int(cl[1])
        if len(cl)>2:
            d.iloc[i,12]=int(cl[2])
    elif int(d.iloc[i,8])>100:
        d.iloc[i,11]=int(d.iloc[i,8])%100
        d.iloc[i,8]=int(d.iloc[i,8])//100
        
```

```python
reasons = d[1].value_counts().to_dict()
reasons.pop(0)

val = list(reasons.values())
key = list(reasons.keys())
plt.figure(figsize=(5,5))
plt.bar(range(len(reasons)),val,tick_label = key,width = 0.5,color='peru')
plt.ylabel("Number of crimes ",fontsize=20)
plt.title("Reason wise ",fontsize=20)
plt.xlabel("Reason",fontsize=20)
#plt.xticks(rotation=90,fontsize=20)
```

```python
d[13] = 0
d[13] = d[2] + d[3]
d[13] = d[13] + d[4]
a = d[13].value_counts().sort_index()
plt.figure(figsize=(5,5))
plt.bar((a.index.astype(int)).astype(str), a.values, color = 'peru', width = 0.4)
plt.xlabel("Number of murders",fontsize=20)
plt.ylabel("Total number of crimes",fontsize=20)
plt.title("Number of people wise",fontsize=20)
plt.show()
```
