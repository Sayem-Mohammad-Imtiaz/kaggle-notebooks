```python
# load data
house_data = pd.read_csv('/kaggle/input/simple-house-price-prediction/house_price.csv')
house_data.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(house_data['area'], house_data['price'])
plt.title('relation between area and price')
plt.xlabel('area')
plt.ylabel('price')
plt.show()
```

```python
plt.scatter(house_data['rooms'], house_data['price'])
plt.title('relation between rooms and price')
plt.xlabel('area')
plt.ylabel('price')
plt.show()
```

```python
x = house_data['area'].values.reshape(-1,1)
y =  house_data['price'].values.reshape(-1,1)
x.shape, y.shape
```

```python
# split the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
# slope and intercept
print(lr.coef_, lr.intercept_)
```

```python
import numpy as np
m, c = lr.coef_, lr.intercept_
plt.scatter(x_train.flatten(), y_train.flatten())
fit_line = [m*x+c for x in x_train.flatten()]
fit_line = np.array(fit_line)
fit_line = fit_line.reshape(-1,1)
plt.plot(x_train.flatten(), fit_line, color='r')
plt.show()
```

```python
pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2score = r2_score(y_test, pred)
mse = mean_squared_error(y_test, pred)
print("R2 score: ",r2score)
print("RMSE: ", np.sqrt(mse))
```
