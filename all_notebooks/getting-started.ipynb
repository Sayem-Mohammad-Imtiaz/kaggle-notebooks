This notebook is meant to get started using Python, we just load the data and take a brief look at it.

```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
pd.set_option('display.max_colwidth', 2000)

import seaborn as sns
```

```python
df_sales = pd.read_csv('../input/superrare/sales.csv')
df_bids = pd.read_csv('../input/superrare/bids.csv')
df_artworks = pd.read_csv('../input/superrare/tokens.csv')
```

```python
# highest sales in USD
df_sales.merge(df_artworks, left_on="tokenId", right_on="tokenId")[["tokenId","eth","rate","usd","name"]].sort_values("usd", ascending=False).head()
```

```python
# distribution of sales (log transform) 
plt.figure()
sns.distplot(np.log(df_sales["usd"].to_list()))
plt.tight_layout()
```

```python
# top artists by sale volume
df_sales.merge(df_artworks, left_on="tokenId", right_on="tokenId")[["tokenId","eth","rate","usd","name","creator"]] \
    .groupby("creator")["usd"] \
    .agg(sum) \
    .sort_values(ascending=False) \
    .plot()
plt.tick_params(
    axis='x',          
    which='both',      
    bottom=False,      
    top=False,         
    labelbottom=False)
plt.yscale('linear') # try 'log'
plt.tight_layout()
```
