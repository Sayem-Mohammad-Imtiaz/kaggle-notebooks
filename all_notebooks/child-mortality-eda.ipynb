```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('-----------------p/ChildMOrtalytRate.csv')
print(df.isnull().sum())

data = df.fillna(method='pad')

print(data.isnull().sum())

df1 = df.rename(columns={'Child Mortality(1 to 4)':'Child Mortality'})
```

```python
countrys = list(dict(list(df1.groupby('Country'))).keys())
n_countrys = len(countrys)
```

```python
for i in range(n_countrys):
    fig,axs = plt.subplots(3,sharex=False,sharey=False,figsize=[10,10])
    fig.suptitle(countrys[i],fontsize=10,color='blue')
    
    data_country = data[data.Country == countrys[i]]

    sns.lineplot(ax=axs[0],x='Year',y='Total Population',hue='Gender',data=data_country)
    sns.lineplot(ax=axs[1],x='Year',y='Child Mortality',hue='Gender',data=data_country)
    sns.lineplot(ax=axs[2],x='Year',y='Mortality Rate',hue='Gender',data=data_country)
    plt.show()
```
