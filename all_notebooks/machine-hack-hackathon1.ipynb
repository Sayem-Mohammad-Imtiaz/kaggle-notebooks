```python
train_raw = pd.read_csv('../input/the-great-indian-hiring-hackathon/Participants_Data_TGIH/Train.csv')
test_raw = pd.read_csv('../input/the-great-indian-hiring-hackathon/Participants_Data_TGIH/Test.csv')
```

```python
train_raw.columns
```

```python
train_temp = train_raw[['StockCode','UnitPrice']]
```

```python
train_temp
```

```python
dictmap = train_temp.groupby(['StockCode']).apply(pd.DataFrame.mode).reset_index(drop=True)
```

```python
test_temp = test_raw[['StockCode']]
```

```python
result = pd.merge(dictmap,test_temp, how='right', on='StockCode')
```

```python
result
```

```python
result.fillna(0, inplace=True)
```

```python
result = result[['UnitPrice']]
```
