```python
Phd_data= pd.read_csv('/kaggle/input/pakistanintellectualcapitalcs/Pakistan Intellectual Capital - Computer Science - Ver 1.csv',encoding="latin1")
```

```python
# Renaming columns names for convinience
phd_data = Phd_data.rename(columns={
    "Teacher Name": "teacher_name",
    "University Currently Teaching": "current_university",
    "Department": "department",
    "Province University Located": "province",
    "Designation": "designation",
    "Terminal Degree": "degree",
    "Graduated from": "graduated_from",
    "Country": "country",
    "Year": "year",
    "Area of Specialization/Research Interests": "specialization",
    "Other Information": "other_information"
})
phd_data.head()  # Displaying initial five rows
```

```python
#this show the the null values in all columns
phd_data.isna().sum()
```

```python
#this code remove the NaN values in specialization columns
df = phd_data.dropna(subset=['specialization'])
```

```python
#the portion represnt the Null values in all columns 
sns.heatmap(df.isnull(), cmap='viridis')
```

```python
df.dtypes
```

```python
df['specialization']
```

```python
# this code will conver all data of specialization column into upper case because it will help in to remove duplication
df['specialization'] = df['specialization'].str.upper()
```

```python
df['specialization']
```

```python
# the code will split the column of the specialization because some teacher participate into many reseach , to count all we need to split data
df = df.drop('specialization', axis=1).join(df['specialization'].str.split('/', expand=True).stack().reset_index(level=1, drop=True).rename('specialization'))
df = df.drop('specialization', axis=1).join(df['specialization'].str.split(',', expand=True).stack().reset_index(level=1, drop=True).rename('specialization'))
df = df.drop('specialization', axis=1).join(df['specialization'].str.split('\n', expand=True).stack().reset_index(level=1, drop=True).rename('specialization'))
df = df.drop('specialization', axis=1).join(df['specialization'].str.split('&', expand=True).stack().reset_index(level=1, drop=True).rename('specialization'))

df.head(10)
```

```python
df["specialization"] = df["specialization"].apply(lambda x: x.replace("DBMS", "DATABASES"))
df["specialization"] = df["specialization"].apply(lambda x: x.replace("DATABASES", "DATA_BASES"))
```

```python
df1 = df["specialization"].value_counts(ascending=False).rename_axis(['specialization']).reset_index(name='counts') #.value_counts()
df1.head(10)
```

```python
px.bar( df1, x= df1.specialization, y='counts', title='Top_10 Books In 2020' , color = 'specialization')
```
