```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datos=pd.read_csv('../input/imdb-data/IMDB-Movie-Data.csv')
df=pd.DataFrame(datos)
```

<p><Left><font color='darkcyan' size = '5'>Gráficas </font></center></p>

<p><Left><font color='slateblue' size = '4'>Votos por Género de Pelicula</font></center></p>

```python
df.groupby("Genre")["Votes"].sum().plot(kind = 'barh', legend ='reverse',figsize=(4,40), color = "aquamarine")
plt.title("Votos por Genero de Pelicula")
```

<p><Left><font color='darkmagenta' size = '3'>La grafica representa cuales fueorn los generos de peliculas más votados desde el 2006 al 2016, teniendo como primer lugar el genero de Acción aventura y ciencia ficción,esta grafica permitira indentificar cuales sonlos generos de pelicula que más le gustan a los espectadores.  </font></center></p>

<p><Left><font color='slateblue' size = '4'>Votos por Año</font></center></p>

```python
df.Votes.groupby(df.Year).sum().plot(kind = 'pie',cmap= 'viridis',figsize=(8,8))
plt.title("Votos por Año")
```

<p><Left><font color='darkmagenta' size = '3'>La grafica de pie nos permite identificar cual fue el año con más votos desde el 2006 al 2016, nos muestra que el numero de votos no varia mucho por año , sin embargo podemos notar una mayor votacion entre los años 2012 y 2014. </font></center></p>

```python
sns.lmplot(x='Year',y= 'Runtime (Minutes)', data=df,fit_reg = False ,hue ="Runtime (Minutes)", legend= False,palette="rocket_r",size = 8)
plt.title("Tiempo de pelicula vs Años ")
```

<p><Left><font color='darkmagenta' size = '3'> Este en un grafico de dispercion en donde podemos observar como el tiempo de la pelicula en minutos va variando con los años, en donde el año con mayor dispercion de datos es el 2016 y con el paso de el tiempo la dispercion va disminuyendo</font></center></p>

<p><Left><font color='slateblue' size = '4'>Relación número de Votos con el Ingreso</font></center></p>

```python
relacion=sns.lmplot(x= 'Revenue (Millions)',y= 'Votes' ,data= df,size=8,line_kws={'color': 'orange'})
plt.title("Relación número de Votos con el Ingreso")
```

<p><Left><font color='darkmagenta' size = '3'> Este en un grafico podemos observar la relacion entre el numero de votos y el ingreso de cada pelicula, en donde podemos además observar una linea de disperción la cla nos indica una relación positiva entre las 2 variables, es decir entre mayor ingreso mayor numero de votos obtenidos</font></center></p>

<p><Left><font color='slateblue' size = '4'>Rating vs Año</font></center></p>

```python
sns.lmplot(x='Year',y= 'Rating', data=df,fit_reg = False ,hue ="Runtime (Minutes)", legend= False,palette="rocket_r",size = 8)
plt.title("Rating vs año")
```

<p><Left><font color='darkmagenta' size = '3'> Este en un grafico podemos observar la relacion entre el numero de votos y el ingreso de cada pelicula, en donde podemos además observar una linea de disperción la cla nos indica una relación positiva entre las 2 variables, es decir entre mayor ingreso mayor numero de votos obtenidos</font></center></p>

<p><Left><font color='slateblue' size = '4'>Tiempo de Pelicula y su Ingreso</font></center></p>

```python
relacion=sns.lmplot(x= 'Revenue (Millions)',y= 'Runtime (Minutes)' ,data= df,size=8, line_kws={'color': 'violet'})
plt.title("Relación entre el Tiempo de Pelicula y su Ingreso")
```

<p><Left><font color='darkmagenta' size = '3'> Este en un grafico podemos observar la relacion entre el tiempo de duración de la pelicula en minutos y el ingreso de la pelicula, la linea de regresión nos indica una relación positiva en donde más ingreso es relacionado con una duracion de pelicula más alta, sin embargo, podemos observar que la mayoria de los datos se situan en un ingreso entre 0 y 200 millones de dolares</font></center></p>
