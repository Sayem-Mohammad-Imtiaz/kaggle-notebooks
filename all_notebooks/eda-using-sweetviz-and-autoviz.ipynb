# Exploratory Data Analysis(EDA)

**EDA** is a technique,widely use for data analysis purpose. EDA is used to get insights out of your data for better model building. EDA is basically intersection of **statistics** and **visualization**.

In this notebook we will do EDA in very simple way. We will do EDA by using two very popular modules

    1. sweetviz
    2. autoviz
   
Both of these modules are very simple to use.

```python
import pandas as pd
import numpy as np
```

```python
car=pd.read_csv("../input/car-pricing-prediction/cars_sampled.csv")
```

```python
car.head(10)
```

```python
car.shape
```

# Sweetviz

**sweetviz is python library used for Exploratory Data Analysis.** It's a very simple to use and easy to understand it's output. 

```python
pip install sweetviz
```

```python
import sweetviz as sv
```

```python
#create a object 
report=sv.analyze(car)
```

**The output of sweetviz library is a html file. We need to provide a filename with html tag in `show_html` function.**

```python
report.show_html("my_first_eda_sweetviz.html")
```

**If you are working on kaggle then you can download this file from kaggle output folder. After downloading this file you can run it on your browser**

# EDA on comparison of two datasets

**We can also perform EDA on comparison of two datasets.** To uderstand this example we will split our dataset in two part

```python
car.shape
```

```python
report_compare = sv.compare(car[:25000],car[25000:])
```

```python
report_compare.show_html("my_first_comparison_eda.html")
```

**The output is same as previous one. Download it and run it on your browser :)**

# EDA Using AutoViz

**From it's name autoviz it is clear that it will do automatic vizualization and do statistics part of EDA for you. Like `sweetviz` it is very easy to use and understand**

```python
pip install autoviz
```

```python
import autoviz
```

```python
from autoviz.AutoViz_Class import AutoViz_Class
```

```python
#run this cell to understand autoviz class

?AutoViz_Class
```

```python
av=AutoViz_Class()
```

```python
autoviz_eda=av.AutoViz('../input/car-pricing-prediction/cars_sampled.csv',verbose=0)
```

**This autoviz generate all possible plots liks kde(kernel density plot),bar chart,scatter etc between all features.**
