```python
import pandas as pd
data = pd.read_csv("../input/game-of-thrones/battles.csv")
data.head(10)
```

```python
data.columns
```

```python
data = data.fillna(0)
data.head(5)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
battlesPerYear = data["year"]
sns.set_theme(style="darkgrid")

sns.countplot(battlesPerYear)

```

```python
import seaborn as sns
plt.figure(figsize = (10,7))
battles = data[["attacker_king",'attacker_outcome']]
battles_new = battles[battles["attacker_king"] != 0 ]
battles_new2 = battles_new[battles_new["attacker_outcome"] != 0]
battles_new2


sns.countplot(x = "attacker_king",hue = "attacker_outcome",data = battles_new2,palette = "Set2")


```

```python
data_char = pd.read_csv("../input/game-of-thrones/character-predictions.csv")
plt.figure(figsize = (10,14))
data_char = data_char.fillna(0)
data_cult = data_char[["culture"]]
data_cult_new = data_cult[data_cult["culture"] != 0]
sns.countplot(y = data_cult_new["culture"],palette ="Set2")

```

```python
data_noble = data_char[["name","isNoble"]]



sns.countplot(x = "isNoble",data = data_noble,palette = "Set2")
plt.xlabel("Noble Characters")
```

```python
char_death = pd.read_csv("../input/game-of-thrones/character-deaths.csv")
char_death = char_death[["Name","Book of Death"]].fillna(0)
char_death_new = char_death[char_death["Book of Death"] != 0]

sns.countplot(y = "Book of Death",data = char_death_new,palette = "Set2")
plt.xlabel("Number of Deaths in each Book")
plt.ylabel("Book Number")
```
