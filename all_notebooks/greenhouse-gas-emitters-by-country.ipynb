```python
import pandas as pd
```

```python
df_ghg = pd.read_csv('/kaggle/input/global-co2-emissions/CO2Emission_LifeExp.csv',index_col=0)
```

```python
(df_ghg[['CO2Emissions']].sort_values(by='CO2Emissions',ascending=False)[0:10]/1000000000).rename(columns={'CO2Emissions': 'CO2 Emissions (GT)'}).round(1)
```

```python
(df_ghg.loc[df_ghg['Population'] >= 10000000,['Population']]/1000000).rename(columns={'Population': 'Population (in millions)'}).join(pd.DataFrame(df_ghg['CO2Emissions']/df_ghg['Population'],columns=['Emissions Per Capita (tons)'])).sort_values(by='Emissions Per Capita (tons)',ascending=False)[0:10].round(1)[['Emissions Per Capita (tons)','Population (in millions)']]
```
