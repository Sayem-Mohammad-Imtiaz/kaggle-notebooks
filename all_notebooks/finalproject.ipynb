```python

dataCleaning= t_emails["text"].tolist()
dataCleaning
```

```python
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

dataCleaning = [each.replace('\r\n',' ' )for each in dataCleaning]
dataCleaning = [each.replace(':',' ' )for each in dataCleaning]
dataCleaning = [each.replace('-',' ' )for each in dataCleaning]
dataCleaning = [each.replace('*',' ' )for each in dataCleaning]
dataCleaning = [each.replace('!',' ' )for each in dataCleaning]
dataCleaning = [each.replace('_',' ' )for each in dataCleaning]
dataCleaning = [each.replace(',',' ' )for each in dataCleaning]
dataCleaning = [each.replace('.',' ' )for each in dataCleaning]
dataCleaning = [each.replace('{',' ' )for each in dataCleaning]
dataCleaning = [each.replace('}',' ' )for each in dataCleaning]
dataCleaning = [each.replace('/',' ' )for each in dataCleaning]
dataCleaning = [each.replace('$',' ' )for each in dataCleaning]
dataCleaning = [each.replace(')',' ' )for each in dataCleaning]
dataCleaning = [each.replace('(',' ' )for each in dataCleaning]
dataCleaning = [each.replace("'",' ' )for each in dataCleaning]
dataCleaning = [each.replace('^',' ' )for each in dataCleaning]
dataCleaning = [each.replace('>',' ' )for each in dataCleaning]
dataCleaning = [each.replace('<',' ' )for each in dataCleaning]
dataCleaning = [each.replace('@',' ' )for each in dataCleaning]
dataCleaning = [each.replace('#',' ' )for each in dataCleaning]
dataCleaning = [each.replace('&',' ' )for each in dataCleaning]
dataCleaning = [each.replace('[',' ' )for each in dataCleaning]
dataCleaning = [each.replace(']',' ' )for each in dataCleaning]
dataCleaning = [each.replace('~',' ' )for each in dataCleaning]
dataCleaning = [each.replace(';',' ' )for each in dataCleaning]
dataCleaning = [each.replace('%',' ' )for each in dataCleaning]
dataCleaning = [each.replace('?',' ' )for each in dataCleaning]
dataCleaning = [each.replace('  ',' ' )for each in dataCleaning]
dataCleaning = [each.replace('=',' ' )for each in dataCleaning]
```

```python
dataCleaning = [each.replace('subject',' ' )for each in dataCleaning]
```

```python

from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
vectorizer.fit(dataCleaning)
vector = vectorizer.transform(dataCleaning)
labels = t_emails['label_num'].tolist()
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()
clf.fit(vector.toarray(), labels)
```
