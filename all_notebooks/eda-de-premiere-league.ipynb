### Visualizacion de los datos

```python
epl.groupby(["Club"])["Goals"].sum()
#Crear un datframe 
Goles_equipos = pd.DataFrame(epl.groupby(["Club"])["Goals"].sum())
```

```python
Goles_equipos
```

```python
import plotly.express as px


fig = px.pie(Goles_equipos, values='Goals', names= Goles_equipos.index, title='Porcentaje de goles por Equipos')
fig.show()
```

De los equipos que descienderon y ascienderon se puede ver que la diferencia de goles entre los tres (Leeds, West Bromwich y Sheffield) es muy favorable para Leeds con un 53.6%

```python
epl.groupby(["Club"])["Passes_Attempted"].sum()
#Crear un datframe 
Pases_equipos = pd.DataFrame(epl.groupby(["Club"])["Passes_Attempted"].sum())
```

```python
Pases_equipos
```

```python
import plotly.express as px


fig = px.pie(Pases_equipos, values='Passes_Attempted', names= Pases_equipos.index, title='Pases por Equipos')
fig.show()
```

Se cree que en el futbol los pases son algo necesario y tienen relacion directa con los goles, por lo que en esta grafica se puede observar que Leeds sigue a la cabeza con un 40.5% de los tres equipos 

```python
import plotly.express as px


fig = px.pie(table_epl, values='W', names='Team', title='Juegos Ganados')
fig.show()
```

```python
import plotly.express as px



fig = px.bar(table_epl, x="Team", y=["W", "D", "L"], title="Total de juegos")
fig.show()
```

La diferencia de puntos entre los 3 es notoria, siendo mas del doble la diferencia de partidos ganados colocando al leed no solo fuera de la zona de descenso si no peleando puestos de media tabla

```python
import plotly.express as px


fig = px.pie(table_epl, values='D', names='Team', title='Juegos Empatados')
fig.show()
```

Se debe de tener un equilibrio para mantener la categoria, por lo que en esta grafica se puede observar que el Leeds mantuvo un equilibrio entre los goles permitidos y los anotados, y en el caso de Sheffiel y West Bromwich se ve clara la diferencia y que perjudica al tener mas goles en contra
