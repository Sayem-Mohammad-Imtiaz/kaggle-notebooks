### What is genotype?

In a broad sense, the term "genotype" refers to the genetic makeup of an organism; in other words, it describes an organism's complete set of genes. In a more narrow sense, the term can be used to refer to the alleles, or variant forms of a gene, that are carried by an organism. Humans are diploid organisms, which means that they have two alleles at each genetic position, or locus, with one allele inherited from each parent. Each pair of alleles represents the genotype of a specific gene.
