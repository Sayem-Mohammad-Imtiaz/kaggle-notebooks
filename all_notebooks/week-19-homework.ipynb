Task: Create a multi-indexed dataframe

Indexes:

level_0 = Sex
level_1 = Embarked
Columns:

level_0 = Sex
level_1 = Survived
Values: Random

Result:

```python
dftc = dft.copy()
dftc.info()

```

```python
index1 = [dftc.Sex.unique(), dftc.Embarked.unique()]
indexes = pd.MultiIndex.from_product(index1, names=["Sex", "Embarked"])
indexes
```

```python
column1 = [dftc.Pclass.unique(), dftc.Survived.unique()]
columns = pd.MultiIndex.from_product(column1, names=["Pclass", "Survived"])
columns
```

```python
dftc1 = pd.DataFrame(np.random.randint(9,size=(8,6)), index=indexes, columns=columns)
dftc1
```

4. Melt:

Task: Create the table below using melt method

Id_vars: 'Name' and 'Embarked'

5. Pivot Table:
Data: https://www.kaggle.com/c/titanic

Task: Create the following pivot tables using the pivot_table method

Index:

A) 'Embarked'

B) 'Embarked', 'Sex', 'Pclass'

C) 'Embarked', 'Sex', 'Pclass'

Values: 'Fare' and 'Age'

aggfunc:

A) 'Fare': mean, 'Age': median

B) 'Fare': mean, 'Age': median

C) 'Fare': mean and np.sum, 'Age': np.median and np.std

```python
#A#
pd.pivot_table(dftc, values=['Fare','Age'], index="Embarked",aggfunc={'Fare':[np.mean,np.sum], 'Age' : [np.median,np.std]})
```

```python
#B#

pd.pivot_table(dftc, values=['Fare','Age'], 
               index=['Embarked', 'Sex', 'Pclass'],
               aggfunc={'Fare':np.mean, 'Age' : np.median})
```

```python
#C#

pd.pivot_table(dftc, values=['Fare','Age'], 
               index=['Embarked', 'Sex', 'Pclass'],
               aggfunc={'Fare':np.mean, 'Age' : np.median})
```
