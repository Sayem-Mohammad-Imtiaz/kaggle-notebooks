```python
# config 
DATASET_DIR = "../input/eurusd-m15/"
SYMBOL = "EURUSD"
```

```python
def get_prices(START, STOP):
    DataName = SYMBOL+"_"+str(PERIOD)+'.csv'
    prices = pd.read_csv(DATASET_DIR + DataName)
    prices['time'] = pd.to_datetime(prices['time'])
    prices = prices.set_index("time")
    prices = prices.rename(columns={'tick_volume':'volume'})
    prices = prices[START < prices.index]
    prices = prices[STOP > prices.index]
    return prices


PERIOD = 60

START_DATE = datetime(2011, 1, 1)
STOP_DATE = datetime(2021, 6, 1)

df = get_prices(START_DATE,STOP_DATE)

print(df.head())

mpf.plot(df,type='line')
```
