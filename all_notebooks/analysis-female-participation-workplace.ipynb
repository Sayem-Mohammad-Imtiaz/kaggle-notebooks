```python
#Add column for Inflation
Inflation_Category = []
for row in df['Inflation_rate']:
    if row < 0:
        Inflation_Category.append('Low')
    elif 0 <= row <= 5:
        Inflation_Category.append('Healthy')
    elif row > 5:
        Inflation_Category.append('High')
    else:
        Inflation_Category.append('Not Defined')
        
df['Inflation_category'] = ""
df['Inflation_category'] = Inflation_Category
df.head(5)
```

```python
#Now, let's create a model, how do level of development and inflation rate determine the level of women participation
#first create a dummy variable for level of development
df['Level_of_development'] = df['Level_of_development'].map({'Developed':1,'Developing':0})

df.tail(5)
```

```python
#use the train,test,split method to predict the model
#import packages
from sklearn import linear_model
import statsmodels.api as sm
from sklearn.model_selection import train_test_split

#define x and y
x = df[['Level_of_development','Inflation_rate']]
y = df['Female_Labor_Force_Participation_rate']

train_x, val_x, train_y, val_y = train_test_split(x,y,random_state=1)

#create the training model
train_model = linear_model.LinearRegression()
train_model.fit(train_x,train_y)

predict_train = train_model.predict(train_x)
print(predict_train)
print("this is the training model")
print("")

#create the validation model
validation_model = linear_model.LinearRegression()
validation_model.fit(val_x,val_y)

predict_validation = validation_model.predict(val_x)
print(predict_validation)
print("this is the validation model")
```

```python
#Calculate R-squared the goodness of fit
from sklearn.metrics import r2_score 

r2_train_model = r2_score(train_y,predict_train)
r2_validation_model = r2_score(val_y,predict_validation)

print(r2_train_model)
print(r2_validation_model)
```

```python
#concluding remarks
print("The model has a r-squared close to zero, implying a very weak effect size. Other variables than level of development and inflation rate contribute to a low or high female labor participation rate. Determinants of female labor participation includes social and cultural factors, accessibility of day care facilities,etc. Note that this data includes only 51 countries; from these countries, you do not know the per capita income, fertility rate, etc.")
```
