```python
import re, nltk
from nltk.corpus import stopwords
stop = stopwords.words('english')

#Removing Punctuations, Numbers, and Special Characters

df2['description'] = df2['description'].str.replace("[^a-zA-Z0-9\s]", " ")
#remove all the words having length 3 or less
df2['description'] = df2['description'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))
#Tokenization: Tokens are individual terms or words, and tokenization is the process of splitting a string of text into tokens.

df2['description']

```

```python
df2["description"] = df2["description"].str.lower().str.split()
df2['description']
```

```python
tokenized_desc = df2['description'].apply(lambda x: [item for item in x if item not in stop])
tokenized_desc
```

```python
#join the tokens back together

for i in range(len(tokenized_desc)):
    tokenized_desc[i] = ' '.join(tokenized_desc[i])

df2['description']=tokenized_desc
df2['description']
```

```python
def transform_format(val):
    if (val == 0).all():
        
        return 255
    else:
        return val
```

```python
cloud_data = ' '.join([line for line in df2.description])
stopword = nltk.corpus.stopwords.words('english')

# plot the WordCloud image                       
wc = WordCloud(background_color='white', max_words=2000, mask=transformed_mask,colormap='Set1')
wc.generate(cloud_data)
wc.to_file('word_cloud.png')
wc.to_image()
```
