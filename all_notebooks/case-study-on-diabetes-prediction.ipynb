```python
# TO PLOT A LINE GRAPH BASED ON PARTICULAR COLUMN WITH THE HELP OF GROUPBY FUNCTION.
import pandas as pd
df=pd.read_csv("../input/diabetes-prediction/Diabetespred.csv")
df1=df.groupby(['DiabetesPedigreeFunction']).count()['Outcome']
sr1=pd.Series(df1)
sr1.plot()
```

```python
# ANOTHER LINE GRAPH BASED ON MEAN OF PREGNANCIES.
import pandas as pd
df=pd.read_csv("../input/diabetes-prediction/Diabetespred.csv")
df1=df.groupby(['Pregnancies']).mean()['BloodPressure']
sr1=pd.Series(df1)
sr1.plot()
```

```python
# TO DISPLAY MULTIPLE LINE GRAPH IN A SINGLE LINE GRAPH BASED ON PARTICULAR COLUMN WE USE THE FOLLOWING.
import pandas as pd 
df=pd.read_csv("../input/diabetes-prediction/Diabetespred.csv")
df1=df.groupby(['Age']).mean()['Pregnancies']
#df1

sr1=pd.Series(df1)
#sr1.plot()


df2=df.groupby(['Age']).mean()['Outcome']
#df1

sr2=pd.Series(df2)
#sr2.plot()

df3=pd.DataFrame({'Pregnancies':sr1,'Outcome':sr2})
df3.plot()
```

```python
# MULTIPLE LINE GRAPH BASICALLY SHOWS THE 2 OR MORE OUTCOMES BASED ON A COMMON GROUP OF INFO.
import pandas as pd 
df=pd.read_csv("../input/diabetes-prediction/Diabetespred.csv")
df1=df.groupby(['Pregnancies']).mean()['Glucose']
#df1

sr1=pd.Series(df1)
#sr1.plot()


df2=df.groupby(['Pregnancies']).mean()['Insulin']
#df1

sr2=pd.Series(df2)
#sr2.plot()

df3=pd.DataFrame({'Glucose':sr1,'Insulin':sr2})
df3.plot()
```
