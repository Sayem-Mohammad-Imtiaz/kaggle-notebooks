```python
#Creating a function to visualize data easily
def plot_covid_data(country,col,plot_ma=False,y_max=200):
 indexer = owid_df[owid_df['location']==country].index
 country_df = owid_df.loc[indexer, 'date' :col]
 country_df = country_df.dropna()
 country_df.set_index('date',inplace=True)
 country_df.drop(country_df.columns.difference([col]), 1,inplace=True)
 country_df.plot(figsize=(12,6), ylim=[0,y_max])
    
 if plot_ma :
    country_df.rolling(window=30).mean()[col].plot()
#We are going to compare Sweden(Did not Impose lockdown) with Pakistan(Impose Lockdown) 
```

# Task-1 Do Lockdown works.?

```python
#We are going to compare Sweden(Did not Impose lockdown) with Pakistan(Impose Lockdown) 
plot_covid_data('Sweden' , 'new_cases_per_million', True)
plot_covid_data('Pakistan' , 'new_cases_per_million', True)
```

### First graph shows sweden(did not impose Lockdown) and second shows Pakistan(Impose lockdown )
### Due to Lockdown,cases in Pakistan Decreses

# Task-2 Does median age effect Death Rate.?

```python
#Here we are going to compare Japan(High Median Age) with India(Low Median Age) 
plot_covid_data('Japan' , 'new_deaths_per_million', True, 2.0)
plot_covid_data('India' , 'new_deaths_per_million', True, 2.0)
```

### Deaths in Low Median Age country like India is high than comparing with  High median age country Like Japan 

```python
#Here we are going to compare United Kingdom with Sweden  because both have nearly same median age
plot_covid_data('Sweden' , 'new_deaths_per_million', True, 30.0)
plot_covid_data('United Kingdom' , 'new_deaths_per_million', True, 30.0)
```

### Countries with nearly same median age have nearly same death rate due to covid-19 

# Task-3 How does Obesity Effect Death Rate.?

```python
#Here we are going to compare two high Obesity countries with Two Low Obesity countries

#High Obesity Rate(US , Canada)
plot_covid_data('United States' , 'new_deaths_per_million', True, 15)
plot_covid_data('Canada' , 'new_deaths_per_million', True, 15)

#Low Obesity Rate(Japan , Bangladesh)
plot_covid_data('Japan' , 'new_deaths_per_million', True, 15)
plot_covid_data('Bangladesh' , 'new_deaths_per_million', True, 15)
```

### Here we analyse that Countries with High Obesity rate have more deaths than countries with Low Obesity Rate

# Task-4 How Does Diabetes effect Death Rate.?

```python
#Here we are going to compare United States(High Diabetes Rate) with Sweden(Low Diabetes Rate)
plot_covid_data('United States' , 'new_deaths_per_million', True, 15)
plot_covid_data('Sweden' , 'new_deaths_per_million', True, 15)
```

### Here we analyse that death rate in countries which have High Diabetes Rate(United States) is above average than countries with Low Diabetes Rate (Sweden). 

# Task-5 Does More Testing Slow Death rate.?

```python
#Here we are going to compare Denmark() with Mexico()

#Denmark has high testing rate
plot_covid_data('Denmark' , 'new_deaths_per_million', True, 15)
#Mexico has low testing rate
plot_covid_data('Mexico' , 'new_deaths_per_million', True, 15)
```

### Here we noticed that providing more testing decreases death rate.
