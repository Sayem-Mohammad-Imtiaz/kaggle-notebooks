# **Importing Libraries**

# **Importing Dataset**

```python
df = pd.read_csv("/kaggle/input/random-salary-data-of-employes-age-wise/Salary_Data.csv")
df.head()
```

# **Checking Correlation**

```python
df.corr()
```

# **HeatMap for Correlation**

```python
sns.heatmap(df.corr(), annot= True)
```

# **Splitting Dependent and Independent Features**

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

# **Splitting Dataset into Training and Testing arrays**

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.05,random_state = 0)
```

# **Training Model**

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit (X_train, y_train)
```

```python
reg.intercept_
```

```python
reg.coef_
```

# **Testing Model**

```python
y_pred = reg.predict(X_test)
y_pred
```

# **R2-Score**

```python
from sklearn.metrics import r2_score
print('r2 Score : ', r2_score(y_test, y_pred))
```

# **Visualizing Training DataSet Result using Matplotlib**

```python
import matplotlib.pyplot as plt 
plt.scatter(X_train, y_train, color = 'cyan')
plt.plot(X_train, reg.predict(X_train),color = 'red')
plt.title('Salary vs Year of experience')
plt.xlabel ('Salary')
plt.ylabel ('YearsExperience')
plt.show()
```
