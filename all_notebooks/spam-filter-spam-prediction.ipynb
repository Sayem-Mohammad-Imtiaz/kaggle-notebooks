```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk import WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
```

```python
nltk.download(['stopwords','punkt','wordnet'])
```

```python
df=pd.read_csv('../input/sms-spam-collection-dataset/spam.csv',encoding = "ISO-8859-1")
df.head()
```

```python
df.shape
```

```python
X=df['v2']
X.head()
```

```python
y=pd.get_dummies(df['v1'])
y=y['spam']
y.head()
```

```python
stemmer=PorterStemmer()
lemmatizer=WordNetLemmatizer()
```

```python
def stemmatize(msg):
  msg=msg.lower()
  msg=re.sub('[^a-zA-Z]',' ',msg)
  msg=[stemmer.stem(word) for word in msg.split() if word not in set(stopwords.words("english"))]
  msg=' '.join(msg)
  return msg
```

```python
def lemmatize(msg):
    msg=msg.lower()
    msg=re.sub('[^a-zA-Z]',' ',msg)
    msg=[lemmatizer.lemmatize(word) for word in msg.split() if word not in set(stopwords.words("english"))]
    msg=' '.join(msg)
    return msg
```

```python
X=df['v2'].apply(lemmatize)
X.head()
```

```python
# Number of unique tokens
ws=set()
for i in X:
  ws.update(set(i.split()))
len(ws)
```

```python
# convert to vectors - Bag of words
bow=CountVectorizer()
X=bow.fit_transform(X.tolist())
```

```python
# convert to vectors - TF-IDF
tfidf=TfidfVectorizer()
X=tfidf.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1)
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

Naive Bayes

```python
nb_clf=MultinomialNB().fit(X_train,y_train)
y_pred=nb_clf.predict(X_test)
```

Naive bayes - Stemmatized - Bag of words

```python
acc_nb_stem_bow = accuracy_score(y_test,y_pred)
print(acc_nb_stem_bow)
print(confusion_matrix(y_test,y_pred))
```

Naive bayes - Lemmatized - Bag of words


```python
acc_nb_lem_bow = accuracy_score(y_test,y_pred)
print(acc_nb_lem_bow)
print(confusion_matrix(y_test,y_pred))
```

Naive bayes - Stemmatized - TF-IDF

```python
acc_nb_stem_tfidf = accuracy_score(y_test,y_pred)
print(acc_nb_stem_tfidf)
print(confusion_matrix(y_test,y_pred))
```

Naive bayes - Lemmatized - TF-IDF

```python
acc_nb_lem_tfidf = accuracy_score(y_test,y_pred)
print(acc_nb_lem_tfidf)
print(confusion_matrix(y_test,y_pred))
```
