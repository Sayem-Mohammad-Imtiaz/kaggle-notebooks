3. Gráfica las distribuciones de cada atributo e indica si el coeficiente de asimetría y curtosis son negativos o positivos.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.distplot(df["hf_score"],hist=False, color="red")
y = np.linspace(0,1,20)
x = np.mean(df["hf_score"])
d1 = np.std(df["hf_score"])
d2 = np.std(df["hf_score"])
plt.plot([x]*20,y)
plt.plot([x-d1]*20,y)
plt.plot([x+d2]*20,y)
plt.plot([x-d1*2]*20,y)
plt.plot([x+d2*2]*20,y)
```
