## Read Data

```python
train = pd.read_csv("../input/video-games-rating-by-esrb/Video_games_esrb_rating.csv")
test = pd.read_csv("../input/video-games-rating-by-esrb/test_esrb.csv")

print(train.head())
```

```python
fig, ax = plt.subplots(figsize=(20,15))
sns.heatmap(data=train.corr(), annot=True)
```

## Split Data

```python
X_train = train.drop(["title","esrb_rating","console"], axis=1)
y_train = train["esrb_rating"]
X_test = test.drop(["title","esrb_rating","console"], axis=1)
y_test = test["esrb_rating"]
```

## Fit Model

#### Decision Tree

```python
model = DecisionTreeClassifier()
model = model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Training Score: ", model.score(X_train, y_train))
print("Test Score: ", model.score(X_test, y_test))

cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)

print(cm)
print(cr)
```

#### Logistic Regression

```python
model = LogisticRegression()
model = model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Training Score: ", model.score(X_train, y_train))
print("Test Score: ", model.score(X_test, y_test))

cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)

print(cm)
print(cr)
```

#### Random Forest

```python
model = RandomForestClassifier()
model = model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Training Score: ", model.score(X_train, y_train))
print("Test Score: ", model.score(X_test, y_test))

cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)

print(cm)
print(cr)
```

#### XGB Classifier

```python
model = XGBClassifier()
model = model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Training Score: ", model.score(X_train, y_train))
print("Test Score: ", model.score(X_test, y_test))

cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)

print(cm)
print(cr)
```

#### Gradent Boosting Classifier

```python
model = GradientBoostingClassifier()
model = model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Training Score: ", model.score(X_train, y_train))
print("Test Score: ", model.score(X_test, y_test))

cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)

print(cm)
print(cr)
```

#### Bagging Classifier

```python
model = BaggingClassifier()
model = model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Training Score: ", model.score(X_train, y_train))
print("Test Score: ", model.score(X_test, y_test))

cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)

print(cm)
print(cr)
```

## Random Forest Testing

```python
model = RandomForestClassifier(n_estimators=200, max_features='log2', criterion='gini', n_jobs=-1)
model = model.fit(X_train, y_train)

y_pred = model.predict(X_test)

print("Training Score: ", model.score(X_train, y_train))
print("Test Score: ", model.score(X_test, y_test))

cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)

print(cm)
print(cr)
```
