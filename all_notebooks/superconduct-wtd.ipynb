```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error
from sklearn.preprocessing import MinMaxScaler
sns.set_style(style='darkgrid')


%matplotlib inline
```

```python
data=pd.read_csv('/kaggle/input/superconduct/superconduct/train.csv')
```

```python
a=['wtd_']
b=['mean','gmean','range','entropy','std']

c=[]
for i in a:
    for j in b:
        c.append(i+j)
print(c)
```

```python
def data_proc(inp):
    a1=[i for i in data.columns if(inp in i)]
    df=data[a1]
    X=df.values
    y=data['critical_temp'].values
  
    return X,y
```

```python
def transformer(x_train,x_test):
    m1=MinMaxScaler()
    x_train=m1.fit_transform(x_train)
    x_test=m1.transform(x_test)
    return x_train,x_test
```

```python
def modeller(x_train,y_train,x_test,num):
    r1=RandomForestRegressor(n_estimators=num,oob_score=True,n_jobs=-1,verbose=1)
    r1.fit(x_train,y_train)
    pred=r1.predict(x_test)
    return pred
```

```python
mae={};mse={}
for k in c:
    X,y=data_proc(k)

    x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)

    x_train,x_test=transformer(x_train,x_test)  
 
    pred=modeller(x_train,y_train,x_test,160)
  
    mae[k]=mean_absolute_error(pred,y_test)
    mse[k]=mean_squared_error(pred,y_test)
```

```python
t1=pd.DataFrame(mae,index=['mae'])
t2=pd.DataFrame(mse,index=['mse'])
```

```python
# plt.figure()

f,axes=plt.subplots(1,2,figsize=(15,6))

axes[0].title.set_text("CRITICAL TEMP : 34.42")
sns.lineplot(data=t1.transpose(),ax=axes[0],markers="H")

sns.lineplot(data=t2.transpose(),ax=axes[1],markers="H")

plt.tight_layout()
```
