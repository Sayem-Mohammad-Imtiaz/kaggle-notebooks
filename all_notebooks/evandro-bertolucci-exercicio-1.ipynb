```python
cat_operador = df.groupby(['aeronave_operador_categoria'],as_index = False).sum().sort_values(by='total_fatalidades', ascending = False)

fig1, ax1 = plt.subplots(figsize=(20, 10))
ax1.pie(cat_operador.total_fatalidades, labels=cat_operador.aeronave_operador_categoria, autopct='%.0f%%',shadow=True, startangle=60)
ax1.axis('equal') 
plt.title('% Total de Fatalidades por Operador')
plt.show()
```

```python
tip_veiculo = df.groupby(['aeronave_tipo_veiculo'],as_index = False).sum().sort_values(by='total_fatalidades', ascending = False)

fig1, ax1 = plt.subplots(figsize=(20, 10))
ax1.pie(tip_veiculo.total_fatalidades,  autopct='%.0f%%',shadow=True, startangle=60)
ax1.axis('equal')
ax1.legend(tip_veiculo.aeronave_tipo_veiculo,
          title="Tipo Veiculo",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))
plt.title('% Total de Fatalidades por Tipo de Veiculo')
plt.show()
```

```python
qntd_motor = df.groupby(['aeronave_motor_quantidade'],as_index = False).sum().sort_values(by='total_fatalidades', ascending = False)

fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes([0,0,1,1])
ax.grid(zorder=0)
ax.set_xlabel('Quantidade de Motor')
ax.set_ylabel('Fatalidades')
ax.bar(qntd_motor.aeronave_motor_quantidade,qntd_motor.total_fatalidades, align='center',zorder=3)
plt.title('Total de Fatalidades por Quantidade de Motor')
plt.show()
```

```python
fase_oper = df.groupby(['aeronave_fase_operacao'],as_index = False).sum().sort_values(by='total_fatalidades', ascending = False)
fase_oper = fase_oper.head(10)
fig = plt.figure(figsize=(20, 5))
ax = fig.add_axes([0,0,1,1])
ax.grid(zorder=0)
ax.set_xlabel('Fase de Operação')
ax.set_ylabel('Fatalidades')
ax.bar(fase_oper.aeronave_fase_operacao,fase_oper.total_fatalidades, align='center',zorder=3)
plt.title('TOP 10 Total de Fatalidades por Fase de Operação')
plt.show()
```

```python
pais_fab = df.groupby(['aeronave_pais_fabricante'],as_index = False).sum().sort_values(by='total_fatalidades', ascending = False)
pais_fab = pais_fab.head(10)
fig = plt.figure(figsize=(20, 5))
ax = fig.add_axes([0,0,1,1])
ax.grid(zorder=0)
ax.set_xlabel('Pais Fabricante')
ax.set_ylabel('Fatalidades')
ax.bar(pais_fab.aeronave_pais_fabricante,pais_fab.total_fatalidades, align='center',zorder=3)
plt.title('TOP 10 Total de Fatalidades por Pais Fabricante')
plt.show()
```

```python
pais_eua = df.loc[df['aeronave_pais_fabricante'] == 'ESTADOS UNIDOS']
pais_eua = pais_eua.groupby(['aeronave_fabricante'],as_index = False).sum().sort_values(by='total_fatalidades', ascending = False)
pais_eua = pais_eua.head(10)
fig = plt.figure(figsize=(20, 5))
ax = fig.add_axes([0,0,1,1])
ax.grid(zorder=0)
ax.set_xlabel('Fabricante')
ax.set_ylabel('Fatalidades')
ax.bar(pais_eua.aeronave_fabricante,pais_eua.total_fatalidades, align='center',zorder=3)
plt.title('TOP 10 Total de Fatalidades por Fabricante nos Estados Unidos')
plt.show()
```
