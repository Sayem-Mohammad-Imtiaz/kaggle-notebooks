```python
import pandas as pd
```

```python
data_train = pd.read_csv('../input/main-tenis/play_tennis_train.csv')
data_test = pd.read_csv('../input/main-tenis/play_tennis_test.csv')

print (data_train.columns.tolist())
```

menampilkan matriks fitur yang terbentuk 
dengan menggunkan library numpy untuk mengubah data menjadi matriks

```python
df = pd.DataFrame(data_train,columns=['outlook','temp','humidity','wind','play'])

fitur_train = pd.get_dummies(df[['outlook','temp','humidity','wind']]).to_numpy()
label_train = df[['play']].to_numpy()

print(fitur_train)
```

membagi data training menjadi 2 :
1. training set (membuat model)
2. validation set/test set (menguji akurasi)

dengan menggunakan library SKLEARN guna membagi data secara acak menjadi training set dan test set

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train,y_test = train_test_split(fitur_train,label_train,random_state=0)
```

training_set digunakan untuk membuat model menggunakan metode linear SVM

```python
from sklearn.svm import LinearSVC #classifier SVM

svc = LinearSVC()
svc.fit(X_train,y_train)
```

model yang terbentuk kemudian diuji ke test_set

```python
#hasil output model akan di simpan dalam variable y_predkemudian dicocokan dengan label yang disimpan dengan variable y_test
y_pred = svc.predict(X_test)
```

menghitung akurasi dataset testing mengunakan confusion matriks

```python
from sklearn.metrics import confusion_matrix

tn, fp, fn, tp = confusion_matrix(y_test,y_pred).ravel()
accuracy = (tn + tp) / (tn+fp+fn+tp)

print (accuracy)
```

```python
df = pd.DataFrame(data_test,columns=['outlook','temp','humidity','wind'])

fitur_test = pd.get_dummies(df[['outlook','temp','humidity','wind']]).to_numpy()

y_pred_test = svc.predict(fitur_test)
print(y_pred_test)
```
