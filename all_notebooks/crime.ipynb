```python
#creating new null cols
df['crime_against_women1']=int(0)
df['crime_against_women2']=int(0)
#splitting col-8 about ',' for diff reasons 
#q is set of i with crime_against cases!=0
q= []
for i in range (0,len(df)):
    r = str(df.iloc[i,8]).split(',')
    
    df.iloc[i,8]=int(r[0])
    if len(r) == 1 and df.iloc[i,8]!=0:
        if df.iloc[i,8] >10:
            df.iloc[i,12]=int(df.iloc[i,8])
            df.iloc[i,8] =int(df.iloc[i,8]%10)
            df.iloc[i,12]=int(df.iloc[i,12]/10)
        q.append(i)
    elif len(r)>1:
        df.iloc[i,12]=int(r[1])
        if len(r)>2:
            df.iloc[i,13] = int(r[2])
print(len(df))
print(len(q))
df.head(5)
```

```python
#mapping for col-8,12,13
#df=df.replace(0, np.nan)
d = {0:np.nan, 1:'Murder with Rape' , 2:'Dowry Deaths', 3:'Suicide', 4:'Kidnapping', 5:'Acid Attack', 6:'Cruelty by Husband/in-laws',7:'Rape only', 8:'Assault on Women with Intent to Outrage her Modesty',9:'Cyber Crimes against Women', 10:'Protection of Children from Sexual Offences Act' }
df['crime_against_women'] = df['crime_against_women'].apply(lambda x: d[x])
df['crime_against_women1'] = df['crime_against_women1'].apply(lambda x: d[x])
df['crime_against_women2'] = df['crime_against_women2'].apply(lambda x: d[x])
df.head(5)
```
