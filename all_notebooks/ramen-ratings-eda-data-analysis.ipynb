## 1) How many different Brands, Variety, Style, and Country?


> Plot different types of Ramen acorss countires.

## 2) What are the Top Ten Ramen? From which countries?

```python
topten = data.dropna()
topten.sort_values(by=['Top Ten'], ascending=False)
topten.head()
```

```python
topten.shape
```

```python
topten.dtypes
```

```python
topten['Top Ten'] = topten['Top Ten'].astype(str)
topten.dtypes
```

```python
def get_year(tt):
  return tt.split('#')[0]
  
def get_rating(tt):
  return tt.split('#')[-1]


topten['Rating'] = topten['Top Ten'].apply(lambda x: f"{get_rating(x)}")
topten['Year'] = topten['Top Ten'].apply(lambda x: f"{get_year(x)}")
topten.head()
```

```python
topten.dtypes
```

```python
plt.figure(figsize=(14,6))
sns.countplot(data=topten, y="Country", order=topten["Country"].value_counts().index, palette='rocket')
plt.xlabel('Count of Top 10 Ramen', fontsize = 10, fontweight= 'bold')
plt.ylabel('Country', fontsize = 10, fontweight = 'bold')
plt.title('Top 10 Ramen Across Countries', fontweight = 'bold', fontsize = 15)
```

Insight: Ramens made from Singapore are most frequently nominated in the Top Ten list, followed by Japan, Malaysia, South Korea and Indonesia.

## 4) Which countries has the most 5 star rated Ramen and What Ramen Style is more likely to be rated 5 stars?

```python
star_float= []

for i in data['Stars']:
  if i == 'Unrated':
    star_float.append(0.00)
  else:
    star_float.append(float(i))

data['Stars']=star_float
data.dtypes
```

```python
fivestar= data[data['Stars']==5]
fivestar
```

```python
fig, ax = plt.subplots(figsize=(15,6))
sns.countplot(x=fivestar['Country'], palette='rocket', order = fivestar['Country'].value_counts().index)
plt.xticks(rotation = 90)
plt.xlabel('Countries', fontsize = 10, fontweight = 'bold')
plt.ylabel('5 Star Rated Ramen', fontsize = 10, fontweight = 'bold')
plt.title('5 Star Rated Ramen Across Countries', fontweight = 'bold', fontsize = 15)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15,6))
sns.countplot(y=fivestar['Style'], order=fivestar['Style'].value_counts().index, palette='rocket')
plt.xlabel('5 Star Rated Ramen', fontsize = 10, fontweight = 'bold')
plt.ylabel('Ramen Styles', fontsize = 10, fontweight = 'bold')
plt.title('5 Star Rated Ramen Styles', fontweight = 'bold', fontsize = 15)
plt.show()
```

Insight: Among 386 ramens rated 5.0 stars,  
> Pack style has the highest frequency. Pack style is more preferred for ramen packaging.


> Japan has the most 5 star rated Ramen, followed by Malaysia

## 5) Which Country has the highest average Stars?

```python
avg_stars=data.groupby('Country').mean().sort_values(by=['Stars'],ascending=False)
avg_stars.head()
```
