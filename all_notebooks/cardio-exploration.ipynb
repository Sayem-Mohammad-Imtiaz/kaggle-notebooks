**Predictive Models**

```python
X_cardio = cardio.drop('cardio',axis=1)
y_cardio = cardio['cardio']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
#Scaling the X data
scaler.fit_transform(X_cardio)
```

```python
# Train Test Split
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_cardio, y_cardio, test_size=0.40, random_state=100)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train, y_train)
```

```python
rfc_cardio_pred = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, rfc_cardio_pred))
print('\n')
print(confusion_matrix(y_test, rfc_cardio_pred))
```

```python
#KNN Prediction
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=2)
```

```python
knn.fit(X_train,y_train)
```

```python
knn_cardio_pred = knn.predict(X_test)
```

```python
print(classification_report(y_test, knn_cardio_pred))
print('\n')
print(confusion_matrix(y_test, knn_cardio_pred))
```

```python
#SVM
from sklearn.svm import SVC
```

```python
model = SVC()
```

```python
model.fit(X_train,y_train)
svc_cardio_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, svc_cardio_pred))
print('\n')
print(confusion_matrix(y_test, svc_cardio_pred))
```

```python
param = {'C': [0.1,1, 10, 100, 1000], 'gamma': [1,0.1,0.01,0.001,0.0001], 'kernel': ['rbf']} 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param,refit=True,verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```
