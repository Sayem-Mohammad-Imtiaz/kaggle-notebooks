```python
import numpy as np
import scipy
from scipy import stats 
import pandas as pd
df=pd.read_csv("/kaggle/input/the-human-freedom-index/hfi_cc_2020.csv")
df.info()
```

```python
cualitativos=[]
cuantitativos=[]
for i in range(len(df.columns)):
    if df.dtypes.values[i]=='object':
        cualitativos.append(df.columns[i])
    else:
        cuantitativos.append(df.columns[i])
        
cuantitativos[1]
```

```python
#Define el rango de cada uno de los atributos (cuantitativos).
#np.max(cuantitativos)- np.min(cuantitativos)
rangos=[]
df.head()
for i in range(len(cuantitativos)):
    rangos.append(max(df[cuantitativos[i]])- min(df[cuantitativos[i]]))
```

```python
 #Obtén la moda o promedio dependiendo el caso para cada atributo.
promedios=[]

for i in range(len(cuantitativos)):
    promedios.append(np.mean( df[cuantitativos[i]] ))
```

```python
#6. Obtén el rango, varianza, desviación estandar para los atributos cuantitativos.
varianzas=[]
desviaciones_estandar=[]
for i in range(len(cuantitativos)):
    varianzas.append(np.var(df[cuantitativos[i]]))
    desviaciones_estandar.append(np.std(df[cuantitativos[i]]))
```

```python
modas=[]
modas2=[]
for i in range(len(cualitativos)):
    modas.append(stats.mode(df[cualitativos[i]]))

for i in range(len(cuantitativos)):
     modas2.append(stats.mode(df[cuantitativos[i]]))
desviaciones_estandar[1]
cuantitativos[1]
```

```python
#Porcentaje de los datos en 1 std
def datos_std1(columna):
    datos_std=[]
    for i in range( len(df[cuantitativos[columna]]) ):
        if (df[cuantitativos[columna]][i] < promedios[columna] + desviaciones_estandar[columna]) & ( df[cuantitativos[columna]][i]> promedios[columna] - desviaciones_estandar[columna] ):
            datos_std.append(df[cuantitativos[columna]][i])
    return datos_std

listas_std=[]
porcentajes_std_columna=[]
for i in range(len(cuantitativos)):
    listas_std.append(datos_std1(i))#datos_std1(i) devuelve los datos de la columna i dentro de istd y se agregan a la lista
    porcentajes_std_columna.append(( len(listas_std[i]) / len(df[cuantitativos[i]])) *100  )
```

```python
porcentajes_std_columna[1]#Aquí están los porcentajes de los datos en 1std de la columna i
```

```python
#Porcentaje de los datos en 2 std
def datos_std2(columna):
    datos_std2=[]
    for i in range( len(df[cuantitativos[columna]]) ):
        if (df[cuantitativos[columna]][i] < promedios[columna] + (2*desviaciones_estandar[columna])) & ( df[cuantitativos[columna]][i]> promedios[columna] - (2*desviaciones_estandar[columna]) ):
            datos_std2.append(df[cuantitativos[columna]][i])
    return datos_std2

listas_std2=[]
porcentajes_std_columna2=[]
for i in range(len(cuantitativos)):
    listas_std2.append(datos_std2(i))#datos_std1(i) devuelve los datos de la columna i dentro de istd y se agregan a la lista
    porcentajes_std_columna2.append(( len(listas_std2[i]) / len(df[cuantitativos[i]])) *100  )
```
