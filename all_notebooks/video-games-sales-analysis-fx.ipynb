# importing the necessary libraries

```python
import numpy as np 
import pandas as pd

import seaborn as sns 
from matplotlib import pyplot as plt
```

# Getting the video games sales data 

```python
df = pd.read_csv('../input/videogamesales/vgsales.csv',index_col='Rank')
df.head()
```

# general info about the data


```python
df.shape
```

```python
df.info()
```

```python
df.describe(include='all')
```

# checking for missing data 

```python
df.isnull().sum()
```

# handling the missing values
# since the number of missing values is small comparing to how much data we have it makes sence to just drop it

```python
df = df.dropna()
df.shape
```

# understanig the data and visualizing it


# platforms 

```python
df['Platform'].value_counts(normalize=True) * 100
```

```python
plt.figure(figsize=(10,8))
df['Platform'].value_counts(normalize=True).plot(kind='pie', fontsize=10,labeldistance=0.85, wedgeprops={'linewidth': 0.8})
plt.legend(loc=0, fontsize=8)
plt.title('platforms pie distrbiution', fontsize=14)
plt.tight_layout()
```

```python
plt.figure(figsize=(8,5))
df['Platform'].value_counts(normalize=True).plot(kind='bar')
plt.title('platforms barplot distrbiution', fontsize=14)
plt.tight_layout()
```

# platform with the most sales


# platforms with most global sales 

```python
platforms = df.groupby(df['Platform'])
```

```python
pgs = platforms['Global_Sales'].sum().sort_values(ascending=False)
top20 = pgs[:20]
```

```python
plt.figure(figsize=(8,5))
top20.plot(kind='bar')
plt.title('top 20 platforms with most global sales', fontsize=14)
plt.tight_layout()
```

# platfroms with most sales (EU)

```python
peus = platforms['EU_Sales'].sum().sort_values(ascending=False)
topEU20 = peus[:20]
```

```python
plt.figure(figsize=(8,5))
topEU20.plot(kind='bar')
plt.title('top 20 platforms with most EU sales', fontsize=14)
plt.tight_layout()
```

# Platforms with most sales (NA)

```python
pnas = platforms['NA_Sales'].sum().sort_values(ascending=False)
topNA20 = pnas[:20]
```

```python
plt.figure(figsize=(8,5))
topNA20.plot(kind='bar')
plt.title('top 20 platforms with most EU sales', fontsize=14)
plt.tight_layout()
```

```python
df.head()
```

# global sales per year

```python
years = df.groupby(df['Year'])
```

# years with most sales

```python
years['Global_Sales'].sum().sort_values(ascending=False)
```

# global sales over time

```python
plt.figure(figsize=(8,5))
years['Global_Sales'].sum().plot(kind='line')
plt.title('global sales over the years', fontsize=14)
plt.grid()
plt.tight_layout()
```

# most popular genre

```python
genres = df.groupby(df['Genre'])
```

```python
genres['Global_Sales'].sum().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,5))
genres['Global_Sales'].sum().sort_values(ascending=False).plot(kind='bar')
plt.title('genre sales (Global) ', fontsize=14)
plt.tight_layout()
```

# correlation between sales

```python
corr = df.corr()
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(corr, cmap='Blues', annot=True)
plt.title('sales correlation', fontsize=14)
plt.tight_layout()
```
