```python
import pandas as pd

import geopandas as gpd
from shapely import wkt

import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (100,15)
```

**Loading all the data as pandas DataFrames**

1. Constituencies data
2. Election result details
3. Geographical data of each constituencies

```python
df_constituencies=pd.read_csv('/kaggle/input/tamil-nadu-2021-state-elections/Tamil_Nadu_State_Elections_2021_Constituency_Metadata.csv')
df_result_details=pd.read_csv('/kaggle/input/tamil-nadu-2021-state-elections/Tamil_Nadu_State_Elections_2021_Details.csv')
df_geo_data=pd.read_csv('/kaggle/input/tamil-nadu-assembly-constituencies-geo-data-2021/constitution-map.csv')
```

Converting pandas dataframe to geo pandas dataframe , please refer here https://geopandas.org/gallery/create_geopandas_from_pandas.html

```python
df_geo_data['geometry'] = df_geo_data['geometry'].apply(wkt.loads)
gdf_geo_data = gpd.GeoDataFrame(df_geo_data, geometry = 'geometry')
```

Now we are merging our geo data with Constituencies meta data , so we have more details of them

```python
gdf_constituency_details=pd.merge(gdf_geo_data,df_constituencies,how='inner',on='Constituency')
gdf_constituency_details.head(2)
```

**Insights map of Total Constituency votes of each(Kind of heat map)**

```python
gdf_constituency_details.plot('Tot_Constituency_votes_polled',legend=True)
```

**Victory Parties - Variation map for winners**

Grouping result details of each constituency so that we can filter out the row of party , based on the votes they got.

```python
df_results_grouped=df_result_details.groupby('Constituency')
```

Lets consider only the EVM Votes for choosing winners

```python
def agg(df):
    return df[df['EVM_Votes']==df['EVM_Votes'].max()]


gpd_const_winners=gpd.GeoDataFrame(pd.merge(df_results_grouped.apply(agg).drop(columns='Constituency'),gdf_constituency_details,on='Constituency'))
# colors_df=pd.DataFrame({'All India Anna Dravida Munnetra Kazhagam':''})
gpd_const_winners.plot('Party',cmap='viridis',legend=True)
```
