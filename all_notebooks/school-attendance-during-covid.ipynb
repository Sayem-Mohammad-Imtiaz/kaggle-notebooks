## Inspect School Level Data

```python
school_level = pd.read_csv("/kaggle/input/school-attendance-in-texas-covid-weather-ses/SchoolLevelData_V3.csv")
school_level.head()
```

```python
school_level.info()
```

## Clean and change column names

```python
school_level.columns
```

```python
school_level = school_level.clean_names()
```

```python
school_level = school_level.remove_empty()
```

```python
school_level.columns
```

```python
school_level = school_level.drop("unnamed_0", axis=1)
school_level = school_level.drop("prcp", axis=1)
school_level = school_level.drop("tmax", axis=1)
school_level = school_level.drop("tmin", axis=1)
```

```python
school_level = school_level.rename_columns({"schooltype": "school_type", "districtnumber": "district_number", "metrostatus": "metro_status", 
                            "attendancepercent": "attendance_percent", "absentpercent": "absent_percent", "covidtotalcountycases": "covid_total_county_cases", 
                            "covidcountypercentincrease": "covid_county_percent_increase", "covidtotalstatecases": "covid_total_state_cases", 
                            "covidstatepercentincrease": "covid_state_percent_increase", "housholdmediumincome": "houshold_medium_income", 
                            "percentwithoutdegree": "percent_without_degree", "percentfamilypoverty": "percent_family_poverty", "unemploymentrate": "unemployment_rate"})
```

```python
school_level.columns
```

## Change data types

```python
school_level.info()
```

```python
school_level['day'] = pd.to_datetime(school_level['day'])
```

```python
school_level['weekday'] = school_level["weekday"].astype("string")
school_level['school'] = school_level["school"].astype("string")
school_level['school_type'] = school_level["school_type"].astype("string")
school_level['district'] = school_level["district"].astype("string")
school_level['county'] = school_level["county"].astype("string")
school_level['tea_description'] = school_level["tea_description"].astype("string")
school_level['nces_description'] = school_level["nces_description"].astype("string")
school_level['metro_status'] = school_level["metro_status"].astype("string")
```
