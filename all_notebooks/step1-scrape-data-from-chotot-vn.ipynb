```python
# get product information on each page
def get_product_one_page():
    # load page source into the program
    page_source = BeautifulSoup(driver.page_source)
    

    name_result = page_source.find_all('h3', class_ = 'adTitle___3SoJh')
    area_result = page_source.find_all('span', class_ = 'adItemCondition___2daw2')
    price_result = page_source.find_all('span', class_ = 'adPriceNormal___puYxd')
    location_result = page_source.find_all('span', class_ = 'item___eld8Q')


    list_names = []
    list_area = []
    list_price = []
    list_location = []
    list_time = []
    
    for item in range(len(name_result)):
        name_item = name_result[item].text
        area_item = area_result[item].text
        price_item = price_result[item].text
        
        list_names.append(name_item)
        list_area.append(area_item)
        list_price.append(price_item)
#         list_time.append(time_item)
        item = item+1
    
    for item in range(len(location_result)):    
        if 'Huyện' in location_result[item].text:
            location_item = location_result[item].text   
            list_location.append(location_item)
        if 'Quận' in location_result[item].text:
            location_item = location_result[item].text   
            list_location.append(location_item)
        item = item+1
        
            
            
    df = pd.DataFrame()
    df['title'] = list_names
    df['area'] = list_area
    df['price_VND'] = list_price
    df['location'] = list_location
#     df['time'] = list_time
        
    return df
```

```python
# get product information on all pages
def get_product_all_pages(number_of_page):
    all_page = pd.DataFrame()
    number_of_pages = number_of_page
    i = 1
    for page in range(number_of_pages):
        print('số trang: '+str(i))
        i=i+1
        one_page = get_product_one_page()
        all_page = all_page.append(one_page)
        sleep(1)
        
#         scroll the page down to the "next" button
        driver.execute_script('window.scrollTo(X=0,Y=4350);')
        next_button = driver.find_element_by_class_name('rightIcon___3o4yy')
        driver.execute_script("arguments[0].click();", next_button)        
    sleep(1)
    return all_page
        
```
