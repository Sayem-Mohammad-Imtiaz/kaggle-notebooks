# Newyork Air Quality

- Columns 
    - **Ozone** 
        - Mean ozone in parts per billion from 1300 to 1500 hours at Roosevelt Island
    - **Solar.R**
        - Solar radiation in Langleys in the frequency band 4000--7700 Angstroms from 0800 to 1200 hours at Central Park
    - **Wind**
        - Average wind speed in miles per hour at 0700 and 1000 hours at LaGuardia Airport
    - **Temp**
        - Maximum daily temperature in degrees Fahrenheit at La Guardia Airport.

```python
fig = plt.figure(figsize=(12,12))
gs = fig.add_gridspec(2,2)

ax0 = fig.add_subplot(gs[0,0])
ax1 = fig.add_subplot(gs[0,1])
ax2 = fig.add_subplot(gs[1,0])
ax3 = fig.add_subplot(gs[1,1])


sns.scatterplot(data=df, ax=ax0,x="Ozone",y="Temp", hue="Temp", palette="icefire")
ax0.set_title("Ozone - Temp")
sns.scatterplot(data=df, ax=ax1,x="Solar.R",y="Temp", hue="Temp", palette="icefire")
ax1.set_title("Solar.R - Temp")
sns.scatterplot(data=df, ax=ax2,x="Wind",y="Temp", hue="Temp", palette="icefire")
ax2.set_title("Wind - Temp")
sns.scatterplot(data=df, ax=ax3,x="Month",y="Temp", hue="Temp", palette="icefire")
ax3.set_title("Month - Temp")

plt.show()
```

## Daily Tempretures by Months

```python
# Daily tempreture
fig = plt.figure(figsize=(20,20))
gs = fig.add_gridspec(3,2)

ax0 = fig.add_subplot(gs[0,0])
ax1 = fig.add_subplot(gs[0,1])
ax2 = fig.add_subplot(gs[1,0])
ax3 = fig.add_subplot(gs[1,1])
ax4 = fig.add_subplot(gs[2,0])
ax5 = fig.add_subplot(gs[2,1])


sns.lineplot(data=df, ax=ax0, x="Day", y=df[df["Month"]==5]["Temp"], linewidth=3, color="turquoise")
ax0.set_title("Month:5 Daily Tempreture")
sns.lineplot(data=df, ax=ax1, x="Day", y=df[df["Month"]==6]["Temp"], linewidth=3)
ax1.set_title("Month:6 Daily Tempreture")
sns.lineplot(data=df, ax=ax2, x="Day", y=df[df["Month"]==7]["Temp"], linewidth=3, color="black")
ax2.set_title("Month:7 Daily Tempreture")
sns.lineplot(data=df, ax=ax3, x="Day", y=df[df["Month"]==8]["Temp"], linewidth=3, color="brown")
ax3.set_title("Month:8 Daily Tempreture")
sns.lineplot(data=df, ax=ax4, x="Day", y=df[df["Month"]==9]["Temp"], linewidth=3, color="orange")
ax4.set_title("Month:9 Daily Tempreture")
sns.lineplot(data=df, ax=ax5, x="Day", y="Temp", hue="Month", palette="icefire")
ax5.set_title("All Month(5-6-7-8-9) Daily Tempreture")
plt.show()
```
