```python
import pandas
import numpy as np
df = pandas.read_csv('../input/mnist-60000-hand-written-number-images/mnist_train.csv',
                    header = None)
```

```python
df.head()
```

```python
print(df.shape)
```

```python
#Features - Pixels values of Images
X = df.drop(0,axis=1).values
#Labels - The labels assigned
Y = df[0].values
```

```python
print(X.shape)
print(Y.shape)
```

```python
im = X[467].reshape(28,28)
```

```python
print(Y[467])
import matplotlib.pyplot as plt
plt.imshow(im)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(X,Y)
```

```python
print(X.shape)
print(xtrain.shape)
```

```python
from sklearn.linear_model import LogisticRegression
lmodel = LogisticRegression()
```

```python
lmodel.fit(xtrain,ytrain)
```

```python
print(lmodel.score(xtrain,ytrain))
print(lmodel.score(xtest,ytest))
```

```python
from sklearn.ensemble import RandomForestClassifier
rmodel = RandomForestClassifier(max_depth=10)
rmodel.fit(xtrain,ytrain)
```

```python
print(xtrain.shape)
```

```python
print(rmodel.score(xtrain,ytrain))
print(rmodel.score(xtest,ytest))
```

```python
print('Logistic Regression Prediction',lmodel.predict([xtest[81]]))
print('Random Forest Prediction',rmodel.predict([xtest[81]]))

print('Actual_Label',ytest[81])

plt.imshow(xtest[81].reshape(28,28))
plt.show()
```

```python
import pandas as pd
dfts = pd.read_csv('../input/mnist-60000-hand-written-number-images/mnist_test.csv',
                  header=None)
dfts.head()
```

```python
Xts = dfts.drop(0,axis=1).values
Yts = dfts[0].values
```

```python
Ypl = lmodel.predict(Xts)
Ypr = rmodel.predict(Xts)
```

```python
print('Accuracy of Logistic model on test csv file',(Yts == Ypl).sum()/len(Xts))
print('Accuracy of Random Forest model on test csv file',(Yts == Ypr).sum()/len(Xts))
```
