```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
sns.set_style('darkgrid')

colors = ['#B90276','#50237F', '#005691', '#008ECF','#E20015', '#00A8B0', '#78BE20', '#006249', '#525F6B']

sns.set_palette(sns.color_palette(colors))
```

# REading the DataSet

```python
data = pd.read_csv("/kaggle/input/cars-dataset-audi-bmw-ford-hyundai-skoda-vw/cars_dataset.csv", sep = ",")
df = pd.DataFrame(data)
df.head()
```

```python
obj1 = df.select_dtypes(include = 'object').columns
obj1  
```

```python
for col in obj1:
    print(col,'\n', df[col].unique(),'\n')
```

```python
float1 = df.select_dtypes(exclude = 'object').columns
float1 
```

# What does the dataset contain

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.countplot(x = 'Make', data = df)
plt.subplot(2,2,2)
sns.countplot(x = 'fuelType', data = df)
plt.subplot(2,2,3)
sns.countplot(x = 'transmission', data = df)
```

# which is the costliest car in the dataset?

```python
plt.figure(figsize = (15,8))
sns.lineplot( y ='price',x = 'year', data = df, hue = 'Make',palette = 'dark' )
```

# Visualizing different makes of Car on Basis of year, tax, mileage, engine size, mpg etc.

```python
i = 1
plt.figure(figsize = (20,20))
for col in float1:
    plt.subplot(3,2,i)
    sns.kdeplot(x = col, data = df, hue = 'Make')
    plt.title(col)
    i+=1
```

```python
df_Audi = df[(df['Make']=='audi')]
```

# Visualizing ONLY Audi models below
***What does Audi dataset contain***

```python
i = 1
plt.figure(figsize = (20,20))
for col in obj1:
    plt.subplot(3,2,i)
    sns.countplot(x = col, data = df_Audi, hue = 'fuelType')
    plt.title(col)
    i+=1
```

# What are the models available in Audi?

```python
df_Audi['model'].unique()
```

```python
i = 1
plt.figure(figsize = (20,20))
for col in float1:
    plt.subplot(3,3,i)
    sns.histplot(x = col, data = df_Audi)
    plt.title(col)
    i+=1
```

```python
df_Audi['fuelType'].unique()
```

# Audi Models comparision with Fuel, Transmission Type & Price
***Audi R8 is the costliest model in the data set***

```python
plt.figure(figsize = (15,15))
plt.subplot(2,1,1)
sns.boxplot(x = df_Audi['model'], y = df_Audi['price'],data = df_Audi, hue = 'fuelType')
plt.title("Audi Models with Price & Fuel Type")
plt.subplot(2,1,2)
sns.boxplot(x = df_Audi['model'], y = df_Audi['price'],data = df_Audi, hue = 'transmission')
plt.title("Audi Models with Price & TRANSMISSION Type")
```

# Sorting Audi Models based on Fuel Types

```python
plt.figure(figsize = (15,7))
df_Audi.groupby(['model', 'fuelType'])[ 'price'].count().plot.bar()
plt.title("Audi Models, Fuel Types & their counts")
```

# Time Vs Price for Audi Models

```python
plt.figure(figsize = (15,8))
sns.lineplot( y ='price',x = 'year', data = df_Audi, hue = 'model',palette = 'dark' )
```

```python
plt.figure(figsize = (20,12))
plt.subplot(2,2,1)
sns.boxplot(x = 'model', y = 'mpg', data = df_Audi)
plt.title("Model Vs MPG")
plt.subplot(2,2,2)
sns.barplot(x = 'model', y = 'price', data = df_Audi)
plt.title("Model Vs Price")
plt.subplot(2,2,3)
sns.barplot(x = 'model', y = 'mileage', data = df_Audi)
plt.title("Model Vs Mileage")
plt.subplot(2,2,4)
sns.boxplot(x = 'model', y = 'tax', data = df_Audi)
plt.title("Model Vs Tax")
```
