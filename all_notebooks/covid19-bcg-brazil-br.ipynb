#Normalization

```python
#1. Lower Case

brazil = brazil.lower()

#2. Punctuation Removal
import string
punc = string.punctuation
brazil = ''.join([char for char in brazil if char not in punc]) #for loops may be inefficient sometimes
```

```python
#1. split()

brazil_words = brazil.split()
brazil_words[:5]
```

```python
#Removing newlines
#strip function does not remove whitespaces from the middle part of the text. It removes white spaces only from the 
#beginnings and end. Let's use replace to get rid of the newlines.

brazil = brazil.replace('\n', ' ')
brazil[:500]
```

```python
# using regular expression to get rid of punctuations:
with open('../input/hackathon/task_1-google_search_txt_files_v2/BR/Brazil-pt-result-29.txt') as g:
    brazil = g.read()

#Normalization
#Lower case

brazil = brazil.lower()

#Removing Punctuation
import re
brazil = re.sub('[^a-zA-Z0-9\s]','',brazil) #replacing not a-z, A-Z, 0-9 and whitespaces with ''.
```

```python
brazil = brazil.replace('\n', ' ')
```

#Tokenization

```python
#1. split()

brazil_words = brazil.split()
brazil_words[:5]
```

```python
#2. nltk

from nltk.tokenize import word_tokenize
brazil_nltk_word =word_tokenize(brazil)
brazil_nltk_word[:2]
```

#Stopwords Removal

```python
from nltk.corpus import stopwords

brazil_stopwords_removed = [w for w in brazil_nltk_word if w not in stopwords.words('english')]
brazil_stopwords_removed[:5]
```

#Parts of Speech Tagging

```python
from nltk import pos_tag
brazil_pos_tag = pos_tag(brazil_stopwords_removed)
brazil_pos_tag[:5]
```

#Named Entity Recognition

```python
from nltk import ne_chunk
brazil_named_entity = ne_chunk(brazil_pos_tag)
```

```python
print(ne_chunk(pos_tag(word_tokenize("coronavirus"))))
```

```python
print(ne_chunk(pos_tag(word_tokenize("brasil"))))
```

#Stemming and Lemmatization

```python
from nltk.stem import LancasterStemmer
from nltk.stem import PorterStemmer

ps = PorterStemmer()
list(map(ps.stem,brazil_stopwords_removed))
```
