```python
df= pd.read_csv('/kaggle/input/oecd-unemployment-rate-2020/2020 unemployment.csv')
df.head()
```

```python
df.isnull().count()
```

```python
df= df.replace([np.inf, -np.inf], np.nan)
```

```python
df.isnull().count()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
def FuncLabelEncoder(df):
    for c in df.columns:
        if df.dtypes[c] == object:
            le.fit(df[c].astype(str))
            df[c] = le.transform(df[c].astype(str))
    return df
```

```python
dfk = FuncLabelEncoder(df)
dfk.info()
```

```python
dfk= dfk.fillna(0)
dfk
```

```python
dfk.drop(['INDICATOR','Flag Codes','MEASURE','FREQUENCY'], axis=1)
```

```python
X= dfk.iloc[:,:-1].values
y= dfk.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2)
```

```python
from xgboost import XGBRegressor
lr= XGBRegressor()
lr.fit(X_train,y_train)
pred= lr.predict(X_test)
from sklearn.metrics import r2_score
print('r2score is ',r2_score(y_test,pred))
```

```python
from sklearn.svm import SVR
lr= SVR()
lr.fit(X_train,y_train)
pred= lr.predict(X_test)
from sklearn.metrics import r2_score
print('r2score is ',r2_score(y_test,pred))
```

```python
from sklearn.linear_model import LinearRegression
lr= LinearRegression()
lr.fit(X_train,y_train)
pred= lr.predict(X_test)
from sklearn.metrics import r2_score
print('r2score is ',r2_score(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
lr= RandomForestRegressor()
lr.fit(X_train,y_train)
pred= lr.predict(X_test)
from sklearn.metrics import r2_score
print('r2score is ',r2_score(y_test,pred))
```

```python
from sklearn.linear_model import Ridge
reg= Ridge()
reg.fit(X_train,y_train)
y_pred= reg.predict(X_test)
from sklearn.metrics import r2_score
print('R2 score is: ',r2_score(y_test,y_pred))
```

```python
from sklearn.neural_network import MLPRegressor
reg = MLPRegressor(random_state=1, max_iter=500).fit(X_train, y_train)
y_pred= reg.predict(X_test)
from sklearn.metrics import r2_score
print('R2 score is: ',r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import AdaBoostRegressor
reg= AdaBoostRegressor()
reg.fit(X_train,y_train)
y_pred= reg.predict(X_test)
from sklearn.metrics import r2_score
print('R2 score is: ',r2_score(y_test,y_pred))  
```

```python
from sklearn.linear_model import SGDRegressor
reg= SGDRegressor()
reg.fit(X_train,y_train)
y_pred= reg.predict(X_test)
from sklearn.metrics import r2_score
print('R2 score is: ',r2_score(y_test,y_pred))  
```
