```python
df1 = df[['Template ID','Call Type']].groupby(['Call Type']).count().reset_index()
df1 = df1.sort_values(['Template ID'], ascending=False)

data = [go.Bar(
            y=df1['Call Type'],
            x=df1['Template ID'],
            orientation = 'h'
    )]


layout = go.Layout(    
    title='Call Type Count',
    margin=dict(
        l=220,
        r=10,
        t=140,
        b=80
    )
)

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

```python
df_ = df[['Template ID','Precinct']]
df_.drop_duplicates(inplace=True)
df2 = df_[['Template ID','Precinct']].groupby(['Precinct']).count().reset_index()
df2 = df2.sort_values(['Template ID'], ascending=False)

data = [go.Bar(
            y=df2['Precinct'][:10],
            x=df2['Template ID'][:10],
            orientation = 'h'
    )]


layout = go.Layout(
    title='Precinct Count (Without duplicate values)',  
    margin=dict(
        l=120,
        r=10,
        t=140,
        b=80
    )
)

fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

**Columns description**

**Template ID**
*Numeric*
Key identifying unique Crisis Templates. This should be used to generate counts.

*WARNING DATA IS DENORMALIZED TO REPRESENT A ONE TO MANY RELATIONSHIP IN "DISPOSITIONS" DO NOT COUNT ROWS WARNING*

**Reported Date**
*String*
Date the Crisis Template was reported in the Records Management System. Can vary +1 day from the occurred date.

**Reported Time**
*String*
Time the Crisis Template was reported in the Records Management System. Can very several hours from the occurred time.

**Occurred Date / Time**
*Date*
Occurred date and time of the event associated with the Crisis Template. Can be null where no association can be made to the underlying event.

**Call Type**
*String*
How the call was received from.

**Initial Call Type**
*String*
Initial classification of the call as made by the communication center at time of dispatch.

**Final Call Type**
*String*
Final classification of the call as made by the officer closing the call. Will differ from the initial call type.

**Disposition**
*String*
Disposition of the Crisis Template (one template per person per crisis involved contact) as reported by the officer.

*WARNING DATA IS DENORMALIZED TO REPRESENT A ONE TO MANY RELATIONSHIP IN "DISPOSITIONS" DO NOT COUNT ROWS WARNING*

**Use of Force Indicator**
*String*
Indicates whether reportable force was used.

**Subject Veteran Indicator**
*String*
Indicates whether the subject of the crisis contact was a identified by the officer as a veteran.

**CIT Officer Requested**
*String*
Indicates whether a Crisis Intervention Team officer was requested to respond to the call.

**CIT Officer Dispatched**
*String*
Indicates whether a Crisis Intervention Team officer was dispatched to the call.

**CIT Officer Arrived**
*String*
Indicates whether a Crisis Intervention Team officer arrived on the call.

**Officer Gender**
*String*
Gender of the officer (self-report), from HR data.

**Officer Race**
*String*
Race of the officer (self-report), from Human Resources (HR) data.

**Officer Year of Birth**
*String*
Year (date part) of the officer’s birth according to HR data.

**Officer Years of Experience**
*String*
Years of law enforcement experience at the time the Crisis Template was reported in the Records Management System.

**CIT Certified Indicator**
*String*
CIT Certification status of the officer at the time of the event.

**Officer Bureau Desc**
*String*
Bureau the officer was assigned to, at the time the Crisis Template was reported in the Records Management System.

**Officer Precinct Desc**
*String*
Precinct the officer was assigned to, at the time the Crisis Template was reported in the Records Management System.

**Officer Squad Desc**
*String*
Squad the officer was assigned to, at the time the Crisis Template was reported in the Records Management System.

**Precinct**
*String*
Geographic precinct area where the call was located. See Precinct and Patrol Boundaries at: [http://www.seattle.gov/police/about-us/about-policing/precinct-and-patrol-boundaries](http://www.seattle.gov/police/about-us/about-policing/precinct-and-patrol-boundaries)

**Sector**
*String*
Geographic sector area where the call was located. See Precinct and Patrol Boundaries at: [http://www.seattle.gov/police/about-us/about-policing/precinct-and-patrol-boundaries](http://www.seattle.gov/police/about-us/about-policing/precinct-and-patrol-boundaries) 

**Beat**
*String*
Geographic beat area where the call was located. See Precinct and Patrol Boundaries at: [http://www.seattle.gov/police/about-us/about-policing/precinct-and-patrol-boundaries](http://www.seattle.gov/police/about-us/about-policing/precinct-and-patrol-boundaries)
