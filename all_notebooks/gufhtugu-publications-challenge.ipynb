# Exploratory analysis

```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

Importing data

```python
df = pd.read_csv('../input/gufhtugu-publications-dataset-challenge/GP Orders - 5.csv', encoding="utf-8", delimiter=',')
#df.dropna(inplace = True)
df.isnull().sum().sort_values(ascending = False)
df[df['Book Name'].isna()]
```

## Top 10 selling books

```python
#Removing NA values
df = df.dropna(how='any',axis=0) 
df = df.assign(Order_Books_Name=df['Book Name'].str.split("/")).explode("Book Name")

book_count = df.Order_Books_Name.explode().value_counts()

#Find Top ten Books
top_ten = book_count[:10]

#Visualization

top_ten.plot.bar()
```

The best-selling book is 'انٹرنیٹ سے پیسہ کمائیں'. Lets try to visualize this data


# **Visualize order status frequency**

```python
counts = df["Order Status"].value_counts()
plt.bar(counts.index, counts.values)
plt.show()
```
