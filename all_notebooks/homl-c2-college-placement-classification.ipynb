### Preprocessing:

```python
initial_df.drop('sl_no', axis=1,inplace=True)
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder = OrdinalEncoder()
gender_encoded = ordinal_encoder.fit_transform(initial_df[['gender']])
initial_df['gender'] = gender_encoded 
ssc_encoded = ordinal_encoder.fit_transform(initial_df[['ssc_b']])
initial_df['ssc_b'] = ssc_encoded
hsc_encoded = ordinal_encoder.fit_transform(initial_df[['hsc_b']])
initial_df['hsc_b'] = hsc_encoded
hsc_s_encoded = ordinal_encoder.fit_transform(initial_df[['hsc_s']])
initial_df['hsc_s'] = hsc_s_encoded
degree_t_encoded = ordinal_encoder.fit_transform(initial_df[['degree_t']])
initial_df['degree_t'] = degree_t_encoded
workex_encoded = ordinal_encoder.fit_transform(initial_df[['workex']])
initial_df['workex'] = workex_encoded
specialisation_encoded = ordinal_encoder.fit_transform(initial_df[['specialisation']])
initial_df['specialisation'] = specialisation_encoded
status_encoded = ordinal_encoder.fit_transform(initial_df[['status']])
initial_df['status'] = status_encoded
initial_df.head()
```

```python
# Scaling data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(initial_df)

# Splitting train test. Although should have been done before EDA, due to small dataset, has been done now 
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(pd.DataFrame(scaled_data), test_size=0.2, random_state=42)
placement_status_trained = train_set[12]
placement_status_test = test_set[12]
train_set.drop([12,13],axis=1,inplace=True)
test_set.drop([12,13],axis=1,inplace=True)
```

```python
from sklearn.pipeline import Pipeline
num_pipeline = Pipeline([
    ('scale', StandardScaler())
])
```

### Only linear test

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(train_set,placement_status_trained)
```

```python
# Testing Linear Regression
some_data = initial_df.drop(['status','salary'],axis = 1)[:5]
same_values = initial_df['status'][:5]
some_data = num_pipeline.fit_transform(some_data)
print (lin_reg.predict(some_data))
print (same_values)
# Works! 
```
