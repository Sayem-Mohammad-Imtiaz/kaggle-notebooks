### SCOPE
Scope hakkında bilmemiz gerekenler:
* global: dosya içerisinde tüm seviyelerde görünür.
* local: tanımladığı fonksiyonda görünür.
* built in scope: 'print', 'len' gibi önceden tanımlanmış yerleşik scope modülündeki isimler.
<br><br>Hadi biraz egzersiz yapalım!!

```python
# tahmin edelim bakalım ne çıkıcak _??
x = 2
def f():
    x = 3
    return x
print(x)      # x = 2 global scope
print(f())    # x = 3 local scope
```

```python
# Peki ya local scope yoksa _??
x = 5
def f():
    y = 2*x        # local bi x yok
    return y
print(f())         # bu yüzden global scope x kullanılır.
# Önce local 'scope' arandı, sonra 'global scope' araştırıldı, eğer hala yoksa son olarak 'built in scope' aranır.
```

<a id="10"></a> <br>
### İÇ İÇE GEÇMİŞ FONKSİYONLAR (NESTED FUNCTIONS)
* fonksiyon içinde fonksiyon
* There is a LEGB rule that is search local scope, enclosing function, global and built in scopes, respectively.
* Saygı ile işleyen LEGB kuralı vardır. Sıra ile local scope, enclosing function, global ve built in scopes aranır.

```python
#iç içe geçmiş fonksiyonlar
def kare():
    """ karenin değerleri döndürülür """
    def ekle():
        """ iki local değişken ekle """
        x = 2
        y = 3
        z = x + y
        return z
    return ekle()**2
print(kare())    
```

<a id="11"></a> <br>
### VARSAYILAN ve ESNEK PARAMETRELER
* Varsayılan parametre örneği:
<br> def f(a, b=1):
        """ b = 1 varsayılan parametredir """
* Esnek parametre örneği:
<br> def f(*args):
       """ *args bir veya daha fazla olabilir"""
<br>def f(** kwargs)
       """ **kwargs ise sözlüktür"""
       
<br><br> hadi biraz da pratik için kod yazalım.

```python
# varsayılan parametreler
def f(a, b = 1, c = 2):
    y = a + b + c
    return y
print(f(5))
# peki varsayılan argümanları değiştirmek istiyorsak
print(f(5,4,3))
```

```python
# esnek parametreler *args
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(1,2,3,4)
# esnek parametreler **kwargs bir sözlüktür
def f(**kwargs):
    """ sözlüğün anahtar ve degerini yazdıralım."""
    for anahtar, deger in kwargs.items():               # Eğer bu bölümü anlamadıysanız döngüler kısmındaki sözlükte döngülere göz atabilirsiniz.
        print(anahtar, " ", deger)
f(ulke = 'ispanya', baskent = 'madrid', nufus = 123456)
```

### LIST COMPREHENSİON
**Bu kernel'daki en önemli konudur.**
<br> Sıklıkla veri analizi için 'list comprehension' kullanıyoruz.
<br> list comprehension: collapse for loops for building lists into a single line
<br> list comprehension: for döngüsü ile yeni bir listeyi tek satırda oluşturma

<br>Örn: num1 = [1,2,3]ve biz bunu şöyle yamak istiyoruz => num2 = [2,3,4].Bu döngü ile yapılabilir. Ancak gereksiz derecede uzundur. Biz bunu 'list comprehension' kullanarak tek satırda yapabiliriz.

```python
# list comprehension örneği
num1 = [1,2,3]
num2 = [i + 1 for i in num1 ]
print(num2)
```

[i + 1 for i in num1 ]: list of comprehension
<br> i +1: list comprehension syntax
<br> for i in num1: for döngüsü syntax
<br> i: yineleyici
<br> num1: yinelenen nesne

```python
# yinelenebilir'deki şartlar(if, else)
num1 = [5,10,15]
num2 = [i**2 if i == 10 else i-5 if i < 7 else i+5 for i in num1]
print(num2)
```

```python
# pokemon csv'ye dönelim ve bir tane daha list comprehension örneği yapalım
# lets classify pokemons whether they have high or low speed. Our threshold is average speed.
# pokemonları yüksek veya düşük hızlara sahip olup olmadığını sınıflayalım. Eşiğimiz ortalama hızdır.
esik = sum(veri.Speed)/len(veri.Speed)
veri["hız_seviyesi"] = ["high" if i > esik else "low" for i in veri.Speed]
veri.loc[:10,["hız_seviyesi","Speed"]] # loc'u daha sonra daha detaylı göreceğiz.
```

<a id="31"></a> <br>
### PANDAS ZAMAN DİZİSİNİ YENİDEN ÖRNEKLEME
* Resampling: Farklı zaman aralıklarında istatistiksel yöntem
* Sıklığı belirtmek için   "M" = month or "A" = year gibi string nesnelere ihtiyacımız var.
* Downsampling: date-time satırlarını günlükten haftalığa doğru biraz daha seyrek frekansa indirme
* Upsampling: date-time satırlarının sıklığı arttırma (günlükten saatliğe doğru)
* Interpolate:  'Linear', "time" veya 'index' gibi farklı yöntemlere göre değerleri düzenleme eksikleri doldurma
    * https://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.interpolate.html


```python
# Geçen bölümde oluşturduğumuz veri2 setini kullanalım.
veri2.resample("A").mean()
```

```python
# Hadi month ile yeniden örneklendirelim.
veri2.resample("M").mean()
# Görebildiğiniz gibi bir çok NaN değeri var çünkü veri2 seti tüm ayları içermiyor.
```

```python
# Gerçek hayatta (veriler gerçek. Biz veri2'yi yaratmamışız sanki) bu problemi interpolate ile çözebiliriz
# İlk değerden interpolate edelim. NaN değerleri dolduralım.
# String ve boolean değerleri doldurmuyor fakat int ve float değerleri interpolate'ın içerdiği farklı yöntemler ile dolduruyor. Örneğin 'linear' olarak
veri2.resample("M").first().interpolate("linear")
```

```python
# birde mean() methodu ile interpolate edelim.
veri2.resample("M").mean().interpolate("linear")
```
