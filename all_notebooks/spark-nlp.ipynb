
### What is Apache Spark ?
Apache Spark is a lightning-fast unified analytics engine for big data and machine learning. It was originally developed at UC Berkeley in 2009.

More details on Spark https://www.kaggle.com/lpdataninja/machine-learning-with-apache-spark

I am using freely available databricks stand alone community edition server (https://community.cloud.databricks.com) as Spark library currently not available in Kaggle directly.
