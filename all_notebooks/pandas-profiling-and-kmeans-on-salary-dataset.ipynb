```python
%%time
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_profiling as pp
%matplotlib inline
```

```python
df = pd.read_csv(r'../input/salary-data-simple-linear-regression/Salary_Data.csv')
df
```

```python
from sklearn.cluster import KMeans
```

```python
distortions= []
for i in range(1,11):
    km = KMeans(n_clusters = i, init = 'k-means++',
               n_init = 10, max_iter = 300)
    km.fit(df)
    distortions.append(km.inertia_)
plt.plot(range(1,11), distortions, marker = 'o')
plt.xlabel('clusters')
plt.ylabel('distortions')
```

```python
km = KMeans(n_clusters = 4, init = 'k-means++',
           n_init = 10, max_iter = 300)
```

```python
km.fit(df)
```

```python
P = km.predict(df)
P
```

```python
df['label'] = P
```

```python
df
```

```python
df0 = df.loc[(df.label) == 0]
df1 = df.loc[(df.label) == 1]
df2 = df.loc[(df.label) == 2]
df3 = df.loc[(df.label) == 3]

plt.scatter(df0.YearsExperience, df0.Salary, color = 'red')
plt.scatter(df1.YearsExperience, df1.Salary, color = 'blue')
plt.scatter(df2.YearsExperience, df2.Salary, color = 'green')
plt.scatter(df3.YearsExperience, df3.Salary, color = 'yellow')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1],color = 'black', marker = '*')
```
