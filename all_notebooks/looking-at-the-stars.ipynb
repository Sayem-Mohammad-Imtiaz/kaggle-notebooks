```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('../input/oec.csv')
df.head()
```

```python
df.describe()
```

```python
df.columns.values
```

```python
disc = df.DiscoveryYear.dropna()
plt.hist(disc, bins=200)
plt.show()
```

# Analyzing DiscoveryYear

```python
disc_pre2000 = disc[disc < 2000]
plt.hist(disc_pre2000, bins=200)
plt.show()
```

```python
disc_post2000 = disc[disc > 2000]
plt.hist(disc_post2000)
plt.show()
```

Wow! The amount of planetary discoveries has increased rampantly in the last few years.
