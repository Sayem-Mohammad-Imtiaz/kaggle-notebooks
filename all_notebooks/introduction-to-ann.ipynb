```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
dataset = pd.read_csv('../input/understanding-ann/Churn_Modelling.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:, 3:-1].values
y = dataset.iloc[:, -1].values
```

```python
print(X)
```

```python
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2])
```

```python
print(X)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

```python
ann = tf.keras.models.Sequential() #input layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu')) #hidden layer 1
ann.add(tf.keras.layers.Dense(units=6, activation='relu')) #hidden layer 2
ann.add(tf.keras.layers.Dense(units=1, activation='sigmoid')) #ouput layer
```

```python
opt = tf.keras.optimizers.Adam(learning_rate=0.01)
ann.compile(optimizer = opt, loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
history = ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```

```python
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Training set'], loc='upper left')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.title('Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Training set'], loc='upper left')
plt.show()
```

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
