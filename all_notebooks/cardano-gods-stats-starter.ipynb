```python
df = pd.read_csv("/kaggle/input/cardano-gods-rarity-stats/cardanogods stats.csv")
df
```

```python
df.info()
```

```python
df["Rarity"] = df["Rarity"].str.replace("%", "")
df["% Sold"] = df["% Sold"].str.replace("%", "")

df["Rarity"] = pd.to_numeric(df["Rarity"])
df["% Sold"] = pd.to_numeric(df["% Sold"])

df.info()
```

# **Rarity of gods**

```python
gods_rarity_mask = df["Class"].str.contains("God")
gods_rarity_df = df[gods_rarity_mask]
gods_rarity_df[:5]
```

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.style as style

style.use("bmh")
```

```python
gods_rarity_df.plot(figsize = (18, 8), kind = "bar", x = "Name/Type", y = "Rarity",
                 color= "gold", edgecolor='black')
plt.title("Gods by rarity")
plt.xlabel("God name")
plt.ylabel("Rarity percentage")
plt.xticks(rotation = 45)
```

# **Rarity of animals(god pets)**

```python
pet_rarity_mask = df["Class"].str.contains("Animal")
pet_rarity_df = df[pet_rarity_mask]
pet_rarity_df[:5]
```

```python
gods_pet_df.plot(figsize = (18, 8), kind = "bar", x = "Name/Type", y = "Rarity",
                 color= "gold", edgecolor='black')
plt.title("Pets by rarity")
plt.xlabel("Pet name")
plt.ylabel("Rarity percentage")
plt.xticks(rotation = 45)
```
