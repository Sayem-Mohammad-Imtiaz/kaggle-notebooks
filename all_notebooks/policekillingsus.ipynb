```python
# Import data
medianHouseHoldInCome = pd.read_csv("../input/fatal-police-shootings-in-the-us/MedianHouseholdIncome2015.csv", encoding = "windows-1252");
percentagePeopleBelowPovertyLevel = pd.read_csv("../input/fatal-police-shootings-in-the-us/PercentagePeopleBelowPovertyLevel.csv", encoding = "windows-1252");
percentOver25CompletedHighSchool = pd.read_csv("../input/fatal-police-shootings-in-the-us/PercentOver25CompletedHighSchool.csv", encoding = "windows-1252");
shareRaceCity = pd.read_csv("../input/fatal-police-shootings-in-the-us/ShareRaceByCity.csv", encoding = "windows-1252");
kill = pd.read_csv("../input/fatal-police-shootings-in-the-us/PoliceKillingsUS.csv", encoding = "windows-1252");
```

```python
# Delete all poverty_rate = "-" items
percentagePeopleBelowPovertyLevel.poverty_rate.replace("-", 0.0, inplace = True);
percentagePeopleBelowPovertyLevel.poverty_rate.value_counts()
```

```python
percentagePeopleBelowPovertyLevel.poverty_rate = percentagePeopleBelowPovertyLevel.poverty_rate.astype("float");
percentagePeopleBelowPovertyLevel.info()
```

```python
# Get poverty rate of each state
areaList = list(percentagePeopleBelowPovertyLevel["Geographic Area"].unique());
print(areaList);
print("The number of states: {}".format(len(percentagePeopleBelowPovertyLevel["Geographic Area"].unique())));
areaPovertyRatio = [];
for each in areaList:
    stateFilter = percentagePeopleBelowPovertyLevel["Geographic Area"] == each
    currentState = percentagePeopleBelowPovertyLevel[stateFilter];
    areaPovertyRate = sum(currentState.poverty_rate) / len(currentState);
    areaPovertyRatio.append(areaPovertyRate);
data = pd.DataFrame({"areaList": areaList, "areaPovertyRatio": areaPovertyRatio});
newIndex = (data["areaPovertyRatio"].sort_values(ascending = False)).index.values; # Sıralanan değerlerin indisleri
sortedData = data.reindex(newIndex);

plt.figure(figsize=(15, 10));
ax = sns.barplot(x = sortedData.areaList, y = sortedData.areaPovertyRatio);
plt.xticks(rotation=90);
plt.xlabel("States");
plt.ylabel("Poverty Rate");
plt.title("Poverty Rata vs States");
```

Percentage of state's population according to races which are black, white, native American, asian and hispanic

```python
#print(shareRaceCity.head());
print(shareRaceCity.info());
```

```python
shareRaceCity.replace("-", 0.0, inplace = True);
shareRaceCity.replace("(X)", 0.0, inplace = True);
shareRaceCity.loc[:, ["share_white", "share_black", "share_native_american", "share_asian", "share_hispanic"]] = shareRaceCity.loc[:, ["share_white", "share_black", "share_native_american", "share_asian", "share_hispanic"]].astype("float");
shareRaceCity.info()
```

```python
areaList = list(shareRaceCity["Geographic area"].unique());
print(areaList);
share_white = [];
share_black = [];
share_native_american = [];
share_asian = [];
share_hispanic = [];

for each in areaList:
    currentAreaFilter = (shareRaceCity["Geographic area"] == each);
    share_white.append(sum(shareRaceCity[currentAreaFilter].share_white) / len(shareRaceCity[currentAreaFilter].share_white));
    share_black.append(sum(shareRaceCity[currentAreaFilter].share_black) / len(shareRaceCity[currentAreaFilter].share_black));
    share_native_american.append(sum(shareRaceCity[currentAreaFilter].share_native_american) / len(shareRaceCity[currentAreaFilter].share_native_american));
    share_asian.append(sum(shareRaceCity[currentAreaFilter].share_asian) / len(shareRaceCity[currentAreaFilter].share_asian));
    share_hispanic.append(sum(shareRaceCity[currentAreaFilter].share_hispanic) / len(shareRaceCity[currentAreaFilter].share_hispanic));
```

```python
# visualization
f,ax = plt.subplots(figsize = (9,15))
sns.barplot(x=share_white,y=areaList,color='green',alpha = 0.5,label='White' )
sns.barplot(x=share_black,y=areaList,color='blue',alpha = 0.7,label='African American')
sns.barplot(x=share_native_american,y=areaList,color='cyan',alpha = 0.6,label='Native American')
sns.barplot(x=share_asian,y=areaList,color='yellow',alpha = 0.6,label='Asian')
sns.barplot(x=share_hispanic,y=areaList,color='red',alpha = 0.6,label='Hispanic')

ax.legend(loc='lower right',frameon = True)     # legendlarin gorunurlugu
ax.set(xlabel='Percentage of Races', ylabel='States',title = "Percentage of State's Population According to Races ")
```
