**We dont need Top# column as we already have index so we drop it**

```python
df.drop('Top #',axis=1,inplace=True)
df
```

**First we need to change the name of coloums 'Twitter ' as it contains one extra space**

```python
df.rename(columns={'Twitter ':'Twitter'},inplace=True)
df
```

```python
df.columns
```

**Now we will make two new columns for account category and subscribers count by splitting the Twitter column**

```python
df.Twitter.str.split(',',expand=True)
```

```python
df[['Name','Category','Subscribers']]=df.Twitter.str.split(',',expand=True)
df
```

**Lets validate these both new columns**

```python
df.isnull().sum()
```

```python
df.Name.unique().size
```

**Now we change Subscribers column type from str to int**

```python
df.replace({
    'Subscribers': '[A-Za-z]',
},'',regex=True,inplace=True)
df['Subscribers'] = pd.to_numeric(df['Subscribers'])
df['Subscribers']
```

```python
calcarr=df['Subscribers'].to_numpy()
calcarr
```

```python
with np.nditer(calcarr, op_flags=['readwrite']) as arr:
    for n in arr:
        if(n<100):
            n[...] = n*1000000
        else:
            n[...] = n*1000    
calcarr
```

```python
df['Subscribers']=pd.Series(calcarr)
df
```

**Now Our DataSet is ready for the analysis !!**

First we groupby the data by category

```python
dff=df.groupby(['Category'])
dff.describe()
```

```python
x=[];
y=[];
dff=df.groupby(['Category'])
for group in dff:
    x.append(group[0])
    y.append(group[1]['Subscribers'].sum())
    print(group[0],end='     Total Subscribers:')
    print(group[1]['Subscribers'].sum())
```

```python
plt.figure(figsize=(28,10))
plt.title('TOTAL SUBCSRIBERS PER CATEGORY')
plt.xlabel('Categories')
plt.ylabel('Subscribers')
plt.bar(x,y)
plt.show()
```

# FORM THE ABOVE CHART IT IS CLEAR THAT PAKISTAN IS MOSTLY WATCHING ENTERTAINMENT VIDEOS AND THEN EDUCATIONAL VIDEOS AND LEAST WACHING VIDEOS IN THE CATEGORY OF 'NONPROFITS & ACTIVISM'.
