Outlier detection

```python
q1, q3= np.percentile(df.Miles, [25,75])
```

```python
iqr = q3 - q1
```

```python
lower_bound = q1 -(1.5 * iqr) 
upper_bound = q3 +(1.5 * iqr)
```

```python
print(lower_bound)
print(upper_bound)
```

```python
sorted(df.Miles)
```

```python
df.Miles[df.Miles >= upper_bound]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```
