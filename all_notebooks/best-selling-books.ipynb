### 1. How many books made it to the list every year?

```python
bestsellers.groupby(["Year"]).size()
```

**So now we know that this list most likely has the Top 50 BestSelling Books for the years 2009 - 2019. Now lets see the split between fiction and non-fiction books**

## 2. What was the split between Fiction/Non-Fiction every year?

```python
bestsellers.groupby(["Year","Genre"]).size()
```

We can see that Non Fiction books in most years make up more than 50% of the BestSeller List

Now Lets Look at

## 3. Which author had the most number of entries on the list?

```python
bestsellers.groupby(["Author","Genre"], as_index=False)["Name"].count().sort_values(by="Name", ascending=False).head(10)
```

## The top 10 Authors Contribute 19% of the entries in the BestSellers List and the Top 5 contribute 11%. Another interesting thing to note is that couple of these authors are not individuals. Lets find out what books they have published

```python
authors = ["American Psychological Association", "Gallup"]
bestsellers.loc[(bestsellers.Author =="American Psychological Association")|(bestsellers.Author =="Gallup")]
```

**Very Interesting that the Publication Manual of the Psychological manual is in the top 10 consistently. Gallup's Strengths finder is also there quite a few times**

Now lets look at

## 4. How has the average price for a book changed across the years?

```python
bestsellers.groupby(["Year"], as_index=False)["Price"].mean()
```

**Surprisingly, the price of books has come down pretty steadily over the years. Makes me wonder if the data started including Kindle prices (which are lower), sometime in the middle.**

Lets also look at this by Genre now

```python
bestsellers[bestsellers["Genre"]=="Non Fiction"].groupby(["Year","Genre"], as_index=False)["Price"].mean()
```

```python
bestsellers[bestsellers["Genre"]=="Fiction"].groupby(["Year","Genre"], as_index=False)["Price"].mean()
```

**Interestingly, the price of Fiction Books went UP till 2014 before starting its climb down. Fiction books have almost always been cheaper than Non Fiction Books**

**Finally, lets look at**

## 5. What is the average number of Reviews received across Genres and how has this changed over time?

```python
bestsellers.groupby(["Year"], as_index=False)["Reviews"].mean()
```

Unsurprisingly, the number of reviews / book has increased steadily over time. 2014 seems interesting however and had almost the same number of reviews as 2019 indicating that there might have been some very highly reviewed books (in terms of # of Reviews) in that year. Lets now see this split by Genre first before investigating 2014

```python
bestsellers[bestsellers["Genre"]=="Fiction"].groupby(["Year","Genre"], as_index=False)["Reviews"].mean()
```

2017 and 2018 were pretty bad years for Fiction Books as the mean number of reviews/book dropped quite a bit before recovering in 2019

```python
bestsellers[bestsellers["Genre"]=="Non Fiction"].groupby(["Year","Genre"], as_index=False)["Reviews"].mean()
```

Fiction Books seem to have a larger number of reviews/book as compared to non-fiction in general. Now lets look at 2014 to see which were the books that contributed to that year being the highest for reviews/book

```python
bestsellers.sort_values(by="Reviews", ascending=False).head(50)
```

So it looks like 2014 had books like Gone Girl and the Fault in our Stars which had a lot of reviews. Another thing we now get to know from this list is that books can appear in the list multiple times if they are listed as bestsellers in multiple yeasr. So lets now find out which books have appeared the most times in this list of bestsellers.

```python
bestsellers.groupby(["Name","Genre"], as_index=False).count().sort_values(by="Author", ascending=False).head(20)
```

The books which appear multiple times i.e in multiple years of the lists are dominated by Non-Fiction books (7/10). Among Fiction, 2 are childrens books. To Kill a MockingBird is the only Fiction book among the top 10 books that have appeared in multiple years of the bestsellers list
