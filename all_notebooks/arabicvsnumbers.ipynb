```python
import os
import numpy as np
import pandas as pd
import random
```

```python
df_arabic=pd.read_csv('../input/ahcd1/csvTrainImages 13440x1024.csv')
df_mnist=pd.read_csv('../input/mnist-digit-recognizer/train.csv')
```

```python
#preproccesing
column_name=[]
for x in range(784):
    column_name.append(x)
df_mnist.drop('label',inplace=True,axis=1)
df_mnist.columns=column_name
clmnm=[]
for x in range(1024):
    clmnm.append(x)
df_arabic.columns=clmnm
clms=[]
for x in range(784,1024):
    clms.append(x)
df_arabic.drop(clms,inplace=True,axis=1)
lstM=['N' for x in range(42000)]
lstA=['A' for x in range(13439)]
df_mnist['label']=lstM
df_arabic['label']=lstA
df=df_mnist.append(df_arabic)
dflabels=df['label']
df=df.drop('label',axis=1)
```

```python
dflabels.shape# this is labels dataframe
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(df, dflabels, test_size=0.10)

```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='liblinear').fit(x_train, y_train)
```

```python
import pickle
pickle.dump(clf,open('logistic','wb'))
```

```python
#to get the score of test cases 
clf.score(x_test,y_test)
```

```python
# it is giving 100% accuracy
```
