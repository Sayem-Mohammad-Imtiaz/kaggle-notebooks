#Applying pseudocolor schemes to image plots

Pseudocolor is a useful tool for enhancing contrast and visualizing data more easily

```python
lum_img = img[:,:,0]
plt.imshow(lum_img)
```

#Color Maps

```python
plt.imshow(lum_img, cmap="hot")
```

```python
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('nipy_spectral')
```

```python
imgplot = plt.imshow(lum_img)
imgplot.set_cmap('summer')
```

```python
imgplot = plt.imshow(lum_img)
plt.colorbar()
imgplot.set_cmap('winter')
plt.title("Color Bar")
```

#Examining a specific data range using histogram
