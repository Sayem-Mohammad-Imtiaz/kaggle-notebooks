Train Test Split & Feature Scaling

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split( df1, y, test_size = 0.25, random_state = 26)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.preprocessing import RobustScaler, StandardScaler 
```

```python
scaler = StandardScaler()
s_train = scaler.fit_transform(x_train)
s_test = scaler.fit_transform(x_test)
```

```python
s_train.shape, s_test.shape
```

Model Selection

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree  import DecisionTreeRegressor
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report , plot_roc_curve
from sklearn.ensemble import RandomForestClassifier
```

```python
lr = LogisticRegression()
tr = DecisionTreeRegressor()
gn  = GaussianNB()
bn  = BernoulliNB()
mn  = MultinomialNB()
rf = RandomForestClassifier(n_estimators = 100)
```

# **Logistic Regression Result**

```python
lr.fit(s_train, y_train)
lrpred = lr.predict(s_test)
lrscore = accuracy_score(y_test, lrpred) 
print('Logistic Regression Accuracy: ', lrscore) 
print(confusion_matrix(y_test, lrpred))
print(classification_report(y_test, lrpred))
plot_roc_curve(lr, s_train, y_train)
plot_roc_curve(lr, s_test, y_test)
```

#  **Decision Tree Result**

```python
tr.fit(s_train, y_train)
trpred = tr.predict(s_test)
trscore = accuracy_score(y_test, trpred) 
print('Decision Tree Accuracy: ', trscore) 
print(confusion_matrix(y_test, trpred))
print(classification_report(y_test, trpred))
```

# **Gaussian NB Result**

```python
gn.fit(s_train, y_train)
gnpred = tr.predict(s_test)
gnscore = accuracy_score(y_test, gnpred) 
print('Gaussian NB Accuracy: ', gnscore) 
print(confusion_matrix(y_test, gnpred))
print(classification_report(y_test, gnpred))
```

# **Bernoulli NB Result**

```python
bn.fit(s_train, y_train)
bnpred = tr.predict(s_test)
bnscore = accuracy_score(y_test, bnpred) 
print('Bernoulli NB Accuracy: ', bnscore) 
print(confusion_matrix(y_test, bnpred))
print(classification_report(y_test, bnpred))
```

# **KNN Result**

```python
kn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', n_jobs = -1)
kn.fit(x_train, y_train)
knpred = tr.predict(x_test)
knscore = accuracy_score(y_test, knpred) 
print('KNN NB Accuracy: ', knscore) 
print(confusion_matrix(y_test, knpred))
print(classification_report(y_test, knpred))
```

# **Multinomial NB**

```python
mn.fit(x_train, y_train)
mnpred = tr.predict(x_test)
mnscore = accuracy_score(y_test, bnpred) 
print('Multinomial NB Accuracy: ', mnscore) 
print(confusion_matrix(y_test, mnpred))
print(classification_report(y_test, mnpred))
```

# **RANDOM FOREST PREDICTION**

```python
rf.fit(s_train, y_train)
rfpred = rf.predict(s_test)
rfscore = accuracy_score(y_test,rfpred) 
print('Random Forest Accuracy: ', rfscore) 
print(confusion_matrix(y_test, rfpred))
print(classification_report(y_test, rfpred))
```
