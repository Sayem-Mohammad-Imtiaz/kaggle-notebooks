# CaféComBytes - Wine Quality

### Carrega datasets para nosso modelo
test.csv será usado para a predição final e para a competição


```python
import pandas as pd
df = pd.read_csv("/kaggle/input/cafecombyterwinequality/train.csv")
dfTestCompetition = pd.read_csv("/kaggle/input/cafecombyterwinequality/test.csv")
```

```python
df
```

Não usamos a coluna de id para treinamento, ela não só representa um identificador do vinho


```python
df = df.drop(columns=["id"])
```

Separa nosso dataset de treino em outros dataset de treino e teste. Para treinar e testar se nosso modelo está bom


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=["quality"]), df["quality"].values, test_size=0.10, random_state=42)
```

É aqui onde nosso modelo é definido e treinado, altere aqui para testar outros modelos

```python
from sklearn.tree import DecisionTreeClassifier
# Cria um modelo de árvore de decisão
model = DecisionTreeClassifier()
# Treino esse modelo, com o método fit. X_train representa as caracteristicas (features) de cada vinho, y_Train é a nota de cada um desses vinhos
model.fit(X_train,y_train)
```

Faz a predição dos dados que não foram usados para o treino do modelo


```python
predictions = model.predict(X_test)
predictions
```

#### Avalia a acurácia do nosso modelo, quanto mais próximo de 1 melhor


```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predictions)
```

Nessa caso, nosso modelo de árvore de decisão conseguiu ter uma acurácia de 72.2%, ou seja, em 72.2% dos casos ela acertou qual era a nota do vinho...

Para submeter as avaliações para a competição temos que usar o dataset de teste para a competição
#### Este dataset não tem a coluna falando qual é a qualidade do vinho, é justamente essa que será enviada para a competição para avaliar os melhores


```python
dfTestCompetition
```

```python
# Novamente, o id não é usado no modelo, ele serve apenas para identificar o vinho, e será útil apenas na hora da submissão
predictionsCompetition = model.predict(dfTestCompetition.drop(columns=["id"]))
predictionsCompetition
```

Criando um dataframe e preenchendo com as informações a serem submetidas

```python
submission = pd.DataFrame()
submission["id"] = dfTestCompetition["id"]
submission["quality"] = predictionsCompetition
```

```python
submission
```

Pra baixar esse dataset e submeter, vamos precisar de uma funçãozinha auxiliar, para gerar um link:

```python
from IPython.display import HTML
import pandas as pd
import numpy as np
import base64

def create_download_link(df, title = "Download CSV file", filename = "data.csv"):  
    csv = df.to_csv(index=False)
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)
```

```python
# Cria um link para baixar nosso resultado em um arquivo .csv
create_download_link(submission)
```
