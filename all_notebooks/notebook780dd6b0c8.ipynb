```python
#import Libraries and data
import pandas as pd
df=pd.read_csv('/kaggle/input/health-care-data-set-on-heart-attack-possibility/heart.csv')
```

```python
#find the mean of data over the target value
mean_df=df.groupby('target').mean()

print(mean_df)
```

```python
'''select the featurs by analyzing the mean value
Here by seeing mean we found that age and fbs are not showing a good impact over the target
Also Cp Showing the most impact
so we don't some of the features for training our model
'''
X= df[['sex', 'cp', 'trestbps', 'chol', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal']]

y= df.target
```

```python
#split the data for traing and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.4)
```

```python
from sklearn.linear_model import LogisticRegression
model= LogisticRegression()
```

```python
#train the model
model.fit(X_train, y_train)

y_pred=model.predict(X_test)
```

```python
#Confusion Metrix for the data

from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```

```python
#Accuracy of model
model.score(X_test, y_test)
```
