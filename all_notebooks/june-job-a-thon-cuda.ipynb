```python
print("Command execution started at:"+str(datetime.datetime.now(IST).replace(microsecond=0).time())+" hrs\n")
 
# Fetching the datetime from the VisitDateTime column and converting it to the standard date and time format
 
k=[]
l=[]
date=[]
diff=[]
prev=[]
for i in range(0,len(data)):
    a=data.VisitDateTime[i]
    if type(data.VisitDateTime[i])==pd.core.scalar._NAType:  # For Nan Values
        w=pandas.NaT
        m=pandas.NaT
        di=pandas.NaT
    elif len(a)==19:
        d=datetime.datetime.fromtimestamp(int(a[:10]))
        w=(d.timestamp())
        m=(d.date())
        di=(today-m).days
    elif len(a)==23:
        d=datetime.datetime.strptime(a,'%Y-%m-%d %H:%M:%S.%f')
        w=(d.timestamp())
        m=(d.date())
        di=(today-m).days
    else:
        pass
    k.append(w)
    l.append(m)
    print("For ",i," of ",len(data))
    print("Visit_Date is: ",m)
    print("Visit_Time is: ",w)
    print("Till_date is: ",di)
    diff.append(di)
    del a,w,m,di
    
 
    dat=data.Signup_Date[i]
    if len(dat)==32:
        a=datetime.datetime.strptime(dat,'%Y-%m-%d %H:%M:%S.%f%z')
    else:
        a=datetime.datetime.strptime(dat,'%Y-%m-%d %H:%M:%S%z')
    dt=a.date()
    date.append(dt)
    vi=today-dt
    print("Signup_Date is: ",dt)
    print("Vintage of the User is:",vi.days,"\n")
    prev.append(vi.days)
    del vi,dt,a

data['Visit_Time']=k
data['Visit_Date']=l
data['Signup_Date']=date
data['Till_date']=diff
data['Vintage']=prev
del k,l,date, prev
data['ProductID']=data['ProductID'].str.upper()
data['OS']=data['OS'].str.upper()
data['Activity']=data['Activity'].str.upper()
 
data=pd.get_dummies(data,columns=['Activity'])
del diff
data.drop(columns=["VisitDateTime"], axis=1, inplace=True)
 
print("\nCommand execution ended at:"+str(datetime.datetime.now(IST).replace(microsecond=0).time())+" hrs")
```
