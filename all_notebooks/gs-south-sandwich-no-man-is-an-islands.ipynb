#“The death of any man reduces me, because I am humanity”. 

By By Natalia Pasternak, 08/15/2020.

The pandemic brings out our inability to see ourselves as part of the whole. The only way to effectively combat Covid-19 is to understand it as a disease of society, much more than of the individual.

And in society, no man - no one - is an island. And no collectivity, either: no family, social or economic sector exists in a bubble, without contact with others. When discussing the resumption of certain activities, it is necessary to keep in mind their articulation with the whole. What happens in pubs affects schools, which in turn affect football games. Interrupting transmission depends much less on isolated measures within each sector, such as rules of behavior for children and athletes, and much more on life outside.

For a restaurant to work, employees need to get to work and leave their children with someone, somewhere (school? Daycare? Brother-in-law?). Soccer players can be tested frequently, but they return home and contaminate their family members. One of them may have children who are schoolmates of the waiter's children. Another, living with a brother who likes to stroll in the mall. The employee who works at a school can become contaminated in the work environment, and, on returning home, on the subway, contaminate other people. To interrupt community transmission, it is necessary to go beyond the navel itself.

For every person who can choose whether to stay at home or go to the mall, there are countless who have no option. If the mall is open, that means that store clerks, managers, security guards and cleaning employees will be forced out of social isolation and using public transport. And for every person who thinks their children will be safe because the school is taking all precautions, there are teachers and staff who have no choice but to expose themselves to come and go as well, and have contact with children whose parents, fool around, went to the mall.
