```python
from pandas_profiling import ProfileReport
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
data=pd.read_csv("../input/men-women-shoes-reviews/Shoes_Data.csv")
```

```python
data['price']=data['price'].str.replace("â‚¹","")
```

```python
data['price']=data['price'].astype(float)
```

```python
data['rating']=data['rating'].str.replace("out of 5 stars","")
```

```python
data['rating']=data['rating'].astype(float)
```

```python
data['total_reviews']=data['total_reviews'].str.replace("ratings","")
```

```python
data['total_reviews']=data['total_reviews'].str.replace("rating","")
```

```python
data['total_reviews']=data['total_reviews'].astype(float)
```
