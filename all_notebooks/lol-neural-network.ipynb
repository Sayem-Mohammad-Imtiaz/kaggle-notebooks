```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
```

```python
 df = pd.read_csv('/kaggle/input/league-of-legends-diamond-ranked-games-10-min/high_diamond_ranked_10min.csv')
```

```python
#display the head of the dataframe
df.head()
```

```python
X = df[df.columns[2:40]]
y = df["blueWins"]
```

```python
y
```

```python
#X for the features, y for the classes.
#80% of the dataset is used for training, 20% left is used to test our model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
clf = MLPClassifier(solver='lbfgs', random_state=1)
#clf = MLPClassifier(solver='adam', activation='relu',alpha=1e-4,hidden_layer_sizes=(50,50,50), random_state=1,verbose=10,learning_rate_init=.1)
clf.fit(X_train,y_train)
MLPClassifier(random_state=1, solver='lbfgs', max_iter=300)
```

```python
predict_train = clf.predict(X_train)
predict_test = clf.predict(X_test)
#performance of the model based on training data
print(confusion_matrix(y_train,predict_train))
print(classification_report(y_train,predict_train))
```

```python
#performance of the model using testing data 
print(confusion_matrix(y_test,predict_test))
print(classification_report(y_test,predict_test))
```

```python
#performance of the whole dataset
predict_dataset = clf.predict(X)
print(classification_report(predict_dataset,y))
my_submission = pd.DataFrame({'gameId': df['gameId'], 'blueWins': predict_dataset})
# you could use any filename. We choose submission here
my_submission.to_csv('submission.csv', index=False)
```
