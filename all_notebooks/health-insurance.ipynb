```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('../input/health-insurance-dataset/Health_insurance.csv')
```

```python
df.describe()
```

Check null values in the data

```python
df.isnull().sum()
```

There are no null values.

Check Outliers

```python
df.plot(kind= 'box')
```

There are outliers in bmi and charges variables. clip() function is used to remove the outliers.

```python
df['bmi']= df['bmi'].clip(upper= df['bmi'].quantile(0.95) )

df['charges']= df['charges'].clip(upper= df['charges'].quantile(0.85) )
```

```python
df.plot(kind= 'box')
```

There are no outliers.

```python
df= df.drop(['region'], axis=1)
```

One Hot Encoding


```python

df= pd.get_dummies(df, columns=['sex', 'smoker'], drop_first= True)
```

```python
df.head()
```

```python
df.columns
```

```python
new_order = ['age', 'bmi', 'children', 'sex_male', 'smoker_yes', 'charges']
```

```python
df= df[new_order]
```

```python
df.head()
```

Correlation

```python
plt.figure(figsize=(8,6))
c= df.corr()
sns.heatmap(c, cmap='coolwarm', annot=True, linewidth=0.25)
plt.xticks(rotation=30)
plt.tight_layout()
```

```python
X= df.iloc[:, :-1]
y= df.iloc[:, -1].values
```

```python
X.head()
```

Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
ss= StandardScaler()
```

```python
cols=['age', 'bmi', 'children']

X[cols]= ss.fit_transform(X[cols])
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=14)
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X_train, y_train)
```

```python
pred= model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

r2_score(y_test, pred)

```

```python
mean_squared_error(y_test, pred)
```
