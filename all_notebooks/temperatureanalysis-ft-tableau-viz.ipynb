## **Data Wrangling**

### **Drop rows where Year in 200,201(typos) and 2020(insufficient data)**

```python
df.drop(df[df["Year"].isin([2020,201,200])].index,inplace=True)

df.describe()
```

#### **Drop rows where AvgTemperature = -99** 

```python
df.drop(df[df["AvgTemperature"]==-99].index,inplace=True)

df.describe()
```

#### **Conversion to Celsius**

```python
df["AvgTemperature"] = round((df["AvgTemperature"]-32)/1.8,1)
```

### **Top 3 "Hottest" Cities of each Region**

```python
df2 = df.groupby(["Region","City","Country"])["AvgTemperature"].mean().round(decimals=2).to_frame()
df2["Rank"] = df2.groupby("Region")["AvgTemperature"].rank(method="first",ascending=False)
df2[df2["Rank"]<=3].sort_values(by=["Region","Rank"])
```

### **Top 3 Coldest Cities of each Region**

```python
df3 = df.groupby(["Region","City","Country"])["AvgTemperature"].mean().round(decimals=2).to_frame()
df3["Rank"] = df3.groupby("Region")["AvgTemperature"].rank(method="first",ascending=True)
df3[df3["Rank"]<=3].sort_values(by=["Region","Rank"])
```

### **Countries most affected by Climate Change**

```python
countries_temp_first = df[df["Year"]==1995].groupby("Country")["AvgTemperature"].mean()
countries_temp_overall = df.groupby("Country")["AvgTemperature"].mean()
countries_first_overall_temps = pd.concat([countries_temp_first.rename("First Temperature"),countries_temp_overall.rename("Overall Temperature")],axis=1).round(decimals=2)
countries_first_overall_temps["Temperature Change"] = abs(countries_first_overall_temps["Overall Temperature"] - countries_first_overall_temps["First Temperature"])
countries_first_overall_temps.sort_values(by="Temperature Change",ascending=False).head(10)
```

### **Cities most affected by Climate Change**

```python
cities_temp_first = df[df["Year"]==1995].groupby(["City","Country"])["AvgTemperature"].mean()
cities_temp_overall = df.groupby(["City","Country"])["AvgTemperature"].mean()
cities_first_overall_temps = pd.concat([cities_temp_first.rename("First Temperature"),cities_temp_overall.rename("Overall Temperature")],axis=1).round(decimals=2)
cities_first_overall_temps["Temperature Change"] = abs(cities_first_overall_temps["Overall Temperature"] - cities_first_overall_temps["First Temperature"])
cities_first_overall_temps.sort_values(by="Temperature Change",ascending=False).head(10)
```
