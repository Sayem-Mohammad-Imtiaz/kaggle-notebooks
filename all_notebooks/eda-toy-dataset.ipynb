```python
# we are import libaray to use project EDA Toy-Dataset
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
# now reading file and looking a head file 
toy = pd.read_csv('../input/toy-dataset/toy_dataset.csv')

toy.head()
```

```python
# Good 
# can cnovert columns names to all lowercase latter 
toy.columns = [x.lower().strip() for x in toy.columns]
toy.columns
```

```python
# Good 
# Knowing the column of the number is it unique or not? To convert it to the Index,
# if it is unique, it can be deleted and not used
toy[toy['number'].duplicated()]['number'].sum()
```

```python
# 0 mean the column is unique 
# set_index 
toy = toy.set_index("number")

toy.head()
```

```python
# looking Data bu Using info()
toy.info()
```

```python
# looking a City Unqiue and Number Unique City
print(toy['city'].unique())
print("Number City Unique is: {}".format(len(toy['city'].unique())))
```

```python
# return  Value_Count of City and What Toy Counts ?
for i in range(len(toy['city'].unique())):
    print("City {} toy number is: {}".format(toy['city'].value_counts().index[i],
                                             toy['city'].value_counts().iloc[i]))
```

```python
# create fig represent City precent in Toy data
values = toy['city'].value_counts().values
label =  toy['city'].value_counts().index
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
plt.pie(values,labels=label, autopct='%.1f%%', explode=[0.05] * len(label))
plt.title("City Percent Represend Toy")
plt.show()
```

```python
# create pie represent gender precent
fig = plt.figure()
ax = fig.add_subplot(1,1,1)

value = toy['gender'].value_counts().values
label = toy['gender'].value_counts().index
plt.pie(value, labels=label,explode= [0.05] * len(label),
        autopct='%0.1f%%')

plt.title("Gender Precent Toy represnt")
plt.show()
```

```python
# create pie represent illness precent

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

value = toy['illness'].value_counts().values
label = toy['illness'].value_counts().index
plt.pie(value, labels=label,explode= [0.05] * len(label),
        autopct='%0.1f%%')

plt.title("illness Precent Toy represnt")
plt.show()
```

```python
# create Counting for illness Value
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
sns.countplot(x="illness", palette="BrBG", data=toy)
plt.xlabel('illness Values')
plt.title("Counting illness Values Sum")
plt.show()
```

```python
# create figure represent city Values
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)
sns.countplot(y="city", palette="husl", data=toy,ax=ax)
plt.ylabel('City Values')
plt.title("Counting City Values")
plt.show()
```

```python
# Create a figure that represents the illness by to gender
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
sns.countplot(x="gender",hue='illness', palette="BrBG", data=toy)
plt.xlabel('gender Values')
plt.title("reprsend gender to illness Values Counts")
plt.legend(loc='best',ncol=2)
plt.show()
```

```python
# Create a figure that represents the City by to gender
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)
sns.countplot(x="gender",hue='city', palette="rocket", data=toy)
plt.legend(loc='best',ncol=2)
plt.show()
```

```python
# Create a figure that represents the Gender by to City

fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)
sns.countplot(x="city",hue='gender', palette="rocket", data=toy)
plt.legend(loc='best')
plt.xticks(rotation=60)
plt.show()
```

```python
# create distplot kde represents age density
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)
sns.distplot(toy['age'],ax=ax)
plt.show()
```

```python
# create distplot kde represents income density
plt.figure(figsize=(10,5))
sns.distplot(toy["income"], color='b')
plt.title("Income distribution")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.distplot(toy[toy["illness"] == "Yes"]["income"], color="g")
plt.title("Income distribution - Illness:Yes")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.distplot(toy[toy["illness"] == "No"]["income"], color="y")
plt.title("Income distribution - Illness:No")
plt.show()
```

```python
# box plot represents income by City
plt.figure(figsize=(10,5))
sns.boxplot(x=toy["income"], y=toy["city"], data=toy)
plt.show()
```
