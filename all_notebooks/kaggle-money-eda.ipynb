```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

dates = pd.read_csv('../input/kaggle-winners/Kaggle_Contest_Dates.csv')
url_dates = list(dates['url'].values)

dates['year'] = pd.to_datetime(dates['date_end']).dt.year

c = dates['year'].value_counts()
c = c.rename_axis('year').reset_index(name='counts')
# print(c)
c.sort_values('year', inplace=True)

# Set the width and height of the figure
plt.figure(figsize=(16,9))

# Add title
plt.title("Number of Kaggle contests with money prizes per year", fontsize=20)

# Bar chart 
sns.barplot(x=c['year'].values, y=c['counts'].values)
plt.tick_params(labelsize=20)

# Add label for vertical axis
plt.ylabel("Number of contests", fontsize=20)
```

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

dates = pd.read_csv('../input/kaggle-winners/Kaggle_Contest_Dates.csv')
url_dates = list(dates['url'].values)

dates['year'] = pd.to_datetime(dates['date_end']).dt.year
groups = dates.groupby('year')

years = []
prizes = []
for year, gr in groups:
    print(year, len(gr))
    prize = gr['prize'].sum()
    years.append(year)
    prizes.append(prize)

c = pd.DataFrame(years, columns=['year'])
c['prize'] = prizes
print(c)
c.sort_values('year', inplace=True)

# Set the width and height of the figure
plt.figure(figsize=(16,9))

# Add title
plt.title("Total money prizes for Kaggle contests (per year).Total: {:.1f}M$".format(dates['prize'].sum() / 1000000), fontsize=20)

# Bar chart 
sns.barplot(x=c['year'].values, y=c['prize'].values)
plt.tick_params(labelsize=20)

# Add label for vertical axis
plt.ylabel("Prize sum for contests (M$)", fontsize=20)

print('Total prize money (overall): {:.1f}M $'.format(dates['prize'].sum() / 1000000))
```

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

dates = pd.read_csv('../input/kaggle-winners/Kaggle_Contest_Dates.csv')
url_dates = list(dates['url'].values)

dates['year'] = pd.to_datetime(dates['date_end']).dt.year
groups = dates.groupby('year')

years = []
prizes_avg = []
for year, gr in groups:
    print(year, len(gr))
    prize = gr['prize'].mean()
    years.append(year)
    prizes_avg.append(prize)

c = pd.DataFrame(years, columns=['year'])
c['prize'] = prizes_avg
print(c)
c.sort_values('year', inplace=True)

# Set the width and height of the figure
plt.figure(figsize=(16,9))

# Add title
plt.title("Average prize for Kaggle contest (per year).", fontsize=20)

# Bar chart 
sns.barplot(x=c['year'].values, y=c['prize'].values)
plt.tick_params(labelsize=20)

# Add label for vertical axis
plt.ylabel("Prize value ($)", fontsize=20)
```
