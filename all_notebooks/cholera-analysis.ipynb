# Data Cleaning

```python
df.isnull().sum()
```

```python
df['Number of reported cases of cholera'] = df['Number of reported cases of cholera'].fillna(method='ffill')
```

```python
df['Number of reported deaths from cholera'] = df['Number of reported deaths from cholera'].fillna(method='ffill')
```

```python
df['Cholera case fatality rate'] = df['Cholera case fatality rate'].fillna(method='ffill')
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.iloc[1059, :]
```

```python
df.iloc[1059, 2] = df.iloc[1059, 2].split(' ')[0]
```

```python
df.iloc[1059, 3] = df.iloc[1059, 3].split(' ')[0]
```

```python
df.iloc[1059, 4] = df.iloc[1059, 4].split(' ')[0]
```

```python
df['Number of reported cases of cholera'] = df['Number of reported cases of cholera'].astype(int)
```

```python
df['Number of reported deaths from cholera'] = df['Number of reported deaths from cholera'].replace('Unknown', 259)
```

```python
df['Number of reported deaths from cholera'] = df['Number of reported deaths from cholera'].astype(int)
```

```python
df['Cholera case fatality rate'] = df['Cholera case fatality rate'].replace('Unknown', 9.64)
```

```python
df['Cholera case fatality rate'] = df['Cholera case fatality rate'].astype(float)
```

```python
df.dtypes
```

# EDA

```python
plt.figure(figsize=(12, 6))

n=1
for i in df.iloc[:, 1:5].columns:
    plt.subplot(2, 2, n)
    sns.distplot(df[i], kde=False, color='r')
    n=n+1
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df)
```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
year = df.Year.value_counts().sort_values(ascending=False).head(20)

plt.figure(figsize=(12, 6))
sns.barplot(year.index, year.values, palette='rocket')
plt.title('Top 20 years with cholera', fontsize=25, weight='bold')
plt.xticks(rotation=90)
plt.show()
```

```python
rep_cases = df.sort_values('Number of reported cases of cholera', ascending=False).head(100)

plt.figure(figsize=(12, 6))
sns.barplot('Number of reported cases of cholera', 'Country', data = rep_cases, palette='pastel')
plt.title('Countries with max cases of cholera', fontsize=25, weight='bold')
#plt.xticks(rotation=90)
plt.show()
```

```python
death_cases = df.sort_values('Number of reported deaths from cholera', ascending=False).head(100)

plt.figure(figsize=(12, 6))
sns.barplot('Country', 'Number of reported deaths from cholera', data = death_cases, palette='Set2')
plt.title('Number of reported deaths cases of cholera', fontsize=25, weight='bold')
plt.xticks(rotation=90)
plt.show()
```

```python
rate = df.sort_values('Cholera case fatality rate', ascending=False).head(100)

plt.figure(figsize=(12, 6))
sns.barplot('Country', 'Cholera case fatality rate', data = rate, palette='mako')
plt.title('Cholera case fatality rate', fontsize=25, weight='bold')
plt.xticks(rotation=90)
plt.show()
```
