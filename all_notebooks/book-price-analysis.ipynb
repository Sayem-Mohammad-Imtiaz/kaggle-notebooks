```python
df = df.drop('SEC Filings',axis=1)
```

```python
print(df.max())
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('Price',axis=1)
y = df['Price']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
X_train['Price/Earnings'] = X_train['Price/Earnings'].astype(float ,errors='raise')
```

```python
X_train['Price/Book'] = X_train['Price/Book'].astype(float ,errors='raise')
```

```python
X_train.info()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
predictions = lr.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
sns.jointplot(x='Price',y='52 Week Low',data=df,kind='reg')
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```
