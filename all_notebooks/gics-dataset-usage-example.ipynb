```python
# Load the data frame
df = pd.read_csv("/kaggle/input/gics-global-industry-classification-standard/gics-map-2018.csv")

df.head()
```

```python
# Get the list of subIndustries with sectorId = 10
df[df["SectorId"] == 10]
```

```python
# Get the list of sectors
sectors = df[["SectorId", "Sector"]].drop_duplicates().set_index("SectorId")
sectors.head(20)
```

```python
# Get the list of industry groups

industry_groups = df[["IndustryGroupId", "IndustryGroup"]].drop_duplicates().set_index("IndustryGroupId")
industry_groups.head(20)
```

```python
# Get the list of industries 

industries = df[["IndustryId", "Industry"]].drop_duplicates().set_index("IndustryId")
industries.head(20)
```

```python
# Get the list of subindustries
industries = df[["SubIndustryId", "SubIndustry", "SubIndustryDescription"]].drop_duplicates().set_index("SubIndustryId")
industries.head(20)
```
