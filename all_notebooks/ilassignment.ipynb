```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

studentinfo_filepath = '../input/ouladdata/studentInfo.csv'
studentinfo_data = pd.read_csv(studentinfo_filepath)

studentinfo_data.tail()

#check type of each descriptive

print(studentinfo_data.dtypes)





```

```python
#check if there are missing values in descriptives

# creating bool series True for NaN values  
bool_series = pd.isnull(studentinfo_data["highest_education"])

# filtering data  
# displaying data only with higher_education = NaN  
studentinfo_data[bool_series]  
```

```python
#Checking if there are some differences in amount of credits taken amongst students based on educational background

analysis_data = pd.read_csv(studentinfo_filepath, index_col="highest_education")
analysis_dataii = pd.read_csv(studentinfo_filepath)
analysis_data

plt.figure(figsize=(20,10))
plt.title("Amount of studied credits based on highest education of student")
sns.barplot(x=analysis_data.index, y=analysis_data['studied_credits'])
plt.ylabel("Studied credits")
```

```python
#Cleaning the original dataset to only contain data from students who have HE Qualification 

x = analysis_dataii.loc[analysis_dataii['highest_education'] == 'HE Qualification']
x.to_csv('cleaned.csv')
```

```python
#Here I will merge the previously filtered, new dataframe with existing studentAssesment.csv in order to find out results of these students

hestudents_filepath = './cleaned.csv'
hestudents_data = pd.read_csv(hestudents_filepath)

studentassesment_filepath = '../input/ouladdata/studentAssessment.csv'
studentassesment_data = pd.read_csv(studentassesment_filepath)
hestudents_data.tail()
print(hestudents_data["id_student"])


y = pd.merge(hestudents_data, studentassesment_data)
y.to_csv('CombinedCSV.csv')
```

```python
#Calculating mean score of all HE students attempting all examinations in studentAssesment.csv

combined_filepath = './CombinedCSV.csv'
combined_data = pd.read_csv(combined_filepath)
combined_data.tail()

combined_data.head()

combined_data.groupby("highest_education")["score"].mean()

```

```python
#Making another cleaned document with only student data of students with low level or no formal education

x = analysis_dataii.loc[analysis_dataii['highest_education'] == 'No Formal quals']
x.to_csv('cleaned2.csv')
```

```python
llstudents_filepath = './cleaned2.csv'
llstudents_data = pd.read_csv(llstudents_filepath)
b = pd.merge(llstudents_data, studentassesment_data)
b.to_csv('CombinedCSV2.csv')
```

```python
#Calculating mean score of all low education or no quals students attempting all examinations in studentAssesment.csv

combined_filepathb = './CombinedCSV2.csv'
combined_datab = pd.read_csv(combined_filepathb)
combined_datab.tail()

combined_datab.head()

combined_datab.groupby("highest_education")["score"].mean()
```

**Conclusion: Students who have HE qualifications score overal higher (by on average 7 points) than students with no previous education/qualification.**

```python
sns.swarmplot(x=combined_datab['score'],
              y=combined_datab['gender'])

```
