```python
dataset = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')
```

```python
new_dataset = pd.DataFrame()
columns = ['type', 'country', 'release_year', 'rating']
for c in columns:
    new_dataset[str(c)] = dataset[c]
new_dataset
```

```python
import random
options = [0,1]
successful = []

for i in range(0,6234,1):
    successful.append(random.choice(options))

new_dataset['successful'] = successful
new_dataset
```

```python
new_dataset.dropna(subset = ["country"], inplace=True)
new_dataset.reset_index(drop=True, inplace=True)
new_dataset
```

```python
for i in range(0,5758,1):
    x = new_dataset.country[i]
    y = x.split(',')
    new_dataset.loc[i,"country"] = len(y)
        
new_dataset = new_dataset.rename(columns={'country' : 'no. of countries'})
```

```python
new_dataset['no. of countries'] = new_dataset['no. of countries'].apply(pd.to_numeric)
new_dataset.info()
```

```python
new_dataset = pd.get_dummies(new_dataset)
new_dataset
```

```python
col = new_dataset.pop('successful')
new_dataset.insert(0, 'successful', col)
new_dataset
```

```python
X = new_dataset.iloc[:, 1:-1].values
y = new_dataset.iloc[:, 0].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=7)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, y_pred))
```
