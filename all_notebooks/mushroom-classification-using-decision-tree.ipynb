Import Libraries

```python
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score,confusion_matrix
```

Check out the Data

```python
mushroom = pd.read_csv("../input/mushroom-classification/mushrooms.csv")
mushroom
```

```python
# It tells that all features are categorical Data
mushroom.info()
```

Check whether null values are present or not

```python
mushroom.isnull().sum()
```

X and y Arrays

```python
X = mushroom.iloc[:,1:]
y = mushroom.iloc[:,0]
```

```python
X
```

```python
y
```

Convert Categorical features using Label Encoder

```python
columns = X.columns

lEncoder = LabelEncoder()
for i in columns:
    X[i] = lEncoder.fit_transform(X[i])
```

```python
X
```

Train Test Split

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,stratify=y)
y.value_counts(normalize=True),y_train.value_counts(normalize=True),y_test.value_counts(normalize=True)
```

Model fitting

```python
clf = DecisionTreeClassifier(criterion="entropy")
clf.fit(X_train,y_train)
```

Plotting Decision Tree

```python
plt.figure(figsize=[15,10])
_ = plot_tree(clf,feature_names=X.columns,filled=True,class_names=["e","p"])
```

Predicting Test Sample

```python
y_predict = clf.predict(X_test)
```

Calculating Accuracy

```python
print(accuracy_score(y_test,y_predict))
print(confusion_matrix(y_test,y_predict))
```
