# Import libraries

```python
%%time
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
```

# Load data set

```python
%%time
df = pd.read_csv('../input/pima-indians-diabetes-database/diabetes.csv')
df.head()
```

# Seperate independent(X) and dependent variables(y)

```python
%%time
X = df.drop('Outcome', axis=1)
X = StandardScaler().fit_transform(X)
y = df['Outcome']
```

# Create train and test data

```python
%%time
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0)
```

# Select Model and assign parameters, here we have chosen SVC , Support vector classifier with cross validation

```python
%%time
model = SVC()

parameters = [{'kernel': ['rbf'], 'gamma': [1e-3, 1e-4],
                     'C': [1, 10, 100, 1000]},
                    {'kernel': ['linear'], 'C': [1, 10, 100, 1000]}]
```

# Fit model using cross validation

```python
%%time
grid = GridSearchCV(estimator=model, param_grid=parameters, cv=5)
grid.fit(X_train, y_train)
```

# Evaluate score by cross-validation

```python
%%time
roc_auc = np.around(np.mean(cross_val_score(grid, X_test, y_test, cv=5, scoring='roc_auc')), decimals=4)
print('Score: {}'.format(roc_auc))
```

# Check Model performance with RandomForestClassifier

```python
%%time
model = RandomForestClassifier(n_estimators=1000)
model.fit(X_train, y_train)
predictions = cross_val_predict(model, X_test, y_test, cv=5)
print(classification_report(y_test, predictions))
```

# Print Confusion Matrix

```python
%%time
print(confusion_matrix(y_test, predictions))
```

# Check cross validation score with RandomForestClassifier 

```python
%%time
score = np.mean(cross_val_score(model, X_test, y_test, cv=5, scoring='roc_auc'))
np.around(score, decimals=4)
```

#Hence we can say that SVC has better score as compared to Random forest classifier
