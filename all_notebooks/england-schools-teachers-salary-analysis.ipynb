# Data Loading

```python
data=pd.read_csv('/kaggle/input/schoolworkforceinengland/data-school-workforce-in-england.csv')
data
```

# Data Munging

```python
data.info()
```

```python
data.columns
```

# Some Coulmns are not necessary in analysis so that's why  drop this columns

```python
data=data.drop(columns={'location','school_type','location_code','geographic_level','age_category'},axis=1)
data
```

**Here one time_period is not in correct form 201920   
0 is not a date so that why i replace it 201921**

```python
data.time_period = data.time_period.replace(201920, 201921)
data
```

***Change the datatype of time_period from integer to datetime***

```python
data['time_period'] = pd.to_datetime(data['time_period'],format='%Y%m%d')
data
```

***Drop the row of a column of Gender which contain total***

```python
data=data[data.gender != 'Total']
```

**Drop the row of a column of Gender which contain Unclassified**

```python
data=data[data.gender != 'Unclassified']
```

**Drop the row of a column of Grade which contain total**

```python
data=data[data.grade !='Total']
```
