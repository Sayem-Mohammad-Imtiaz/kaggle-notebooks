```python
df = pd.read_csv("../input/movie-synopsis-dataset/movie_synopsis.csv")
df.head(10)
```

```python
df.isnull().sum()
```

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from string import punctuation
stopwords = set(stopwords.words('english'))
import string
print(stopwords)
```

```python
df2 = df[['synopsis']]
```

```python
df2['synopsis'] = df2['synopsis'].str.lower().str.split()
df2['synopsis'] =df2['synopsis'].apply(lambda x: [item for item in x if item not in stopwords])
df2['synopsis'] = df2['synopsis'].apply(lambda x : [item.translate(str.maketrans('', '', string.punctuation)) for item in x])
```

```python
df2['synopsis']
```

```python
df2explode=pd.DataFrame(df2['synopsis'].explode())
df2explode
```

```python
df2explode.value_counts()
```

```python
df2wordcount=pd.DataFrame(df2explode.groupby('synopsis')['synopsis'].count().sort_values(ascending=False))
```

```python
df2wordcount
```

```python
df2wordcount=df2wordcount.rename(columns={'synopsis': 'count'})
df2wordcount.index = df2wordcount.index.rename('Word')
df2wordcount =df2wordcount.reset_index()
df2wordcount
```

```python
count400 = df2wordcount[df2wordcount['count']>400]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

count400.plot.bar(x='Word')
```

```python
import seaborn as sns
sns.barplot(y = 'Word', x = 'count', data = count400)
```

```python
count800 = df2wordcount[df2wordcount['count']>800]
sns.barplot(y = 'Word', x = 'count', data = count800)
```
