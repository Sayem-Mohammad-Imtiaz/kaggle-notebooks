# ****Data preprocessing

In the Area column, there are two values that we need to care about: the area of the house and the number of bedrooms in the house. So we separate the Area column into 2 different columns for easy handling.

```python

df1[['area_m2', 'number_of_bedrooms']] = df1['area'].str.split('-',expand=True)
```

```python
df1 = df1.drop(['area','Unnamed: 0'],1)
```

So, we replace 'tỷ' to number

```python
for index,row in df1.iterrows():
    if len(str(row.price_VND)) == 8:
        row['price_VND'] = row['price_VND'].replace(" tỷ",",000,000")
    if len(str(row.price_VND)) == 7:
        row['price_VND'] = row['price_VND'].replace(" tỷ","0,000,000")
    if len(str(row.price_VND)) == 6:
        row['price_VND'] = row['price_VND'].replace(" tỷ","00,000,000")
    if len(str(row.price_VND)) == 4:
        row['price_VND'] = row['price_VND'].replace(" tỷ",",000,000,000")
    if 'triệu' in row.price_VND:
        row['price_VND'] = row['price_VND'].replace(" triệu",",000,000")
# for index,row in df1.iterrows():
#     row['price'] = row['price'].replace(",","")
```

```python
for index,row in df1.iterrows():
    row['price_VND'] = row['price_VND'].replace(" tỷ",",000,000,000")
```
