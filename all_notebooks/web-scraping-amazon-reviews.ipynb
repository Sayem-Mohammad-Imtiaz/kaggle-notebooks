```python
df= pd.read_csv('../input/amazon-manual-coffe-grinder-review/Kaggle_Data_Coffee.csv')
```

```python
df
```

```python
def getdate(column):
    return column.split('on')[-1]
    
```

```python
def getlocation(column):
    return column.split('on')[0]
    
```

```python
df['Location']= df['Date'].apply(getlocation)
df['Date']=df['Date'].apply(getdate)

```

```python
df
```

```python
def Country(text):
    return text.split('the')[-1]
```

```python
df['Country']=df['Location'].apply(Country)
```

```python
df.drop('Location',axis=1,inplace=True)
```

```python
df
```

```python
len(df['Date'][0])
```

```python
df= df[['Username','Country','Date','Review','Rating']]
```

```python
df
```

```python
df['Date'].dtype
```

```python
df['Date']=pd.to_datetime(df['Date'], errors='coerce')
```

```python
df['Review_Length']= df['Review'].apply(len)
```

```python
df= df[['Username','Country','Date','Review','Review_Length','Rating']]
```

```python
df
```

```python
def getstars(fromtext):
    return fromtext[:1]
```

```python
df['Stars/5']=df['Rating'].apply(getstars)
```

```python
df.drop('Rating',axis=1,inplace=True)
```

```python
df
```

```python
df=df.astype({'Stars/5': 'int64'})
```

```python
df['Stars/5'].mean()
```

```python
plt.figure(figsize=(10,4))
plt.title('Stars(Out of five), Mean=3.96')
sns.countplot(x='Stars/5',data=df,palette='PuRd')
```

```python
plt.figure(figsize=(10,4))
plt.xlim([datetime.date(2017, 1, 1), datetime.date(2021, 7, 19)])
sns.kdeplot(x='Date',data=df,color='purple',fill=True)
```

```python
plt.figure(figsize=(10,5))
plt.xlim([datetime.date(2016, 8, 1), datetime.date(2021, 7, 19)])
plt.yticks([0,1,2,3,4,5])
sns.kdeplot(x='Date',y='Stars/5',data=df,fill=True,levels=6,color='purple')

plt.figure(figsize=(10,5))
plt.xlim([datetime.date(2016, 8, 1), datetime.date(2021, 7, 19)])
sns.kdeplot(x='Date',hue='Stars/5',palette='BuPu',fill=True,data=df)
```

```python
plt.figure(figsize=(10,4))
sns.regplot(x='Review_Length',y='Stars/5',color='purple',data=df)
```
