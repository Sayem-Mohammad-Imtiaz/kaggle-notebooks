```python
df = pd.read_csv('../input/usa-covid19-vaccinations/us_state_vaccinations.csv')
```

```python
df.head()
```

Because the question asks state data specifically, I decided to remove entities that didn't make up the 50 states

```python
df['location'].unique()
entities_to_discard = ['Bureau of Prisons', 'Dept of Defense', 'Long Term Care', 'Veterans Health', 'Federated States of Micronesia', 'Guam', 'Indian Health Svc','Marshall Islands', 'Northern Mariana Islands', 'Republic of Palau',
                      'Virgin Islands', 'Puerto Rico', 'American Samoa', 'District of Columbia', 'United States']
df = df[~df['location'].isin(entities_to_discard)] #select data that does NOT fall into the above list
```

```python
df.info()
```

N/A values indicate the data is missing for a particular column. Since we're working with numeric data, I set missing values to 0.

```python

df.fillna(0, inplace=True)
```

Next, I needed to convert the 'date' feature into datetime format in order to select the data from last week's timeframe

```python

df['date'] = pd.to_datetime(df['date'])
todays_date = df['date'].max()
start_week = todays_date -  pd.to_timedelta('6 day')

df_7day = df[df['date'] >= start_week] #selecting the data from this past week
```

## List the top 10 states that distributed the most vaccines total this past week

```python

df_7day.groupby('location')['total_distributed'].sum().sort_values(ascending=False).head(10)
```

## Give the top 10 states who distributed the most vaccines per capita this past week

```python

df_7day.groupby('location')['distributed_per_hundred'].sum().sort_values(ascending=False).head(10)
```
