 A ride-sharing company is investigating complaints by its drivers that some of
the passengers (traveling with children) do not conform with child safety guidelines (for
example, not bringing a child seat or not using the seat belt). The company undertakes
surveys in two major cities. The surveys are collected independently, with one sample
being taken from each city. From the data collected, it seems that the passengers in City
B are more noncompliant than those in City A. The law enforcement authority wants to
know if the proportion of passengers conforming with child safety guidelines is different
for the two cities. The data for the two cities is given in the following table:
City A City B
Total surveyed 200 230
Number of people compliant 110 106
Conduct the appropriate test, at 5% significance level, to test the hypothesis.

(p1-p2)/sqrt(pc(1-pc)/N1 +pc(1-pc)/N2)

```python
p1=110/200
p2=106/230
n1=200
n2=230
pc=(110+106)/(n1+n2)
z=(p1-p2)/(((pc*(1-pc)/n1)+(pc*(1-pc/n2)))**0.5)
stats.norm.cdf(z)
p=2*(1-stats.norm.cdf(z))
p
```

```python
from scipy.stats import pearsonr
data1 = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
data2 = [0.353, 3.517, 0.125, -7.545, -0.555, -1.536, 3.350, -1.578, -3.537, -1.579]
stat, p = pearsonr(data1, data2)
print('stat=%2.10f, p=%2.10f' % (stat, p))
if p > 0.05:
	print('Probably independent')
else:
	print('Probably dependent')
```

```python
from scipy.stats import spearmanr
data1 = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
data2 = [0.353, 3.517, 0.125, -7.545, -0.555, -1.536, 3.350, -1.578, -3.537, -1.579]
stat, p = spearmanr(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably independent')
else:
	print('Probably dependent')
```

```python
from scipy.stats import ttest_ind
data1 = [0.873, 2.817, 0.121, -0.945, -0.055, -1.436, 0.360, -1.478, -1.637, -1.869]
data2 = [1.142, -0.432, -0.938, -0.729, -0.846, -0.157, 0.500, 1.183, -1.075, -0.169]
stat, p = ttest_ind(data1, data2)
print('stat=%.3f, p=%.3f' % (stat, p))
if p > 0.05:
	print('Probably the same distribution')
else:
	print('Probably different distributions')
```

```python
def avg(n):
    x=n.split(",")
    c=0
    ni=0
    for i in x:
        if "*" in i:
            c=c+int(i[0:-1])
        else:
            c=int(i)+c
            ni=ni+1
    return (c/ni)
              
```

```python
import statsmodels as sm
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
plt.plot(y)
```

```python
plot_acf(y,lags=100,alpha=0.05);
plot_pacf(y,lags=100,alpha=0.05);
```

## Moving Averages

```python
ar=np.array([1])
ma=np.array([1,0.8])
arma_process=ArmaProcess(ar,ma)
m=arma_process.generate_sample(nsample=1000)
```

```python
plt.plot(m)
```

$$ y_t=\alpha +\beta *\epsilon_t $$

```python
plot_acf(m);
plot_pacf(m);
```

```python
ar=np.array([1])
ma=np.array([1,0.5,0.3,0.1])
arma_process=ArmaProcess(ar,ma)
m=arma_process.generate_sample(nsample=1000)
```

```python
plt.plot(m)
```

$$ y_t=\alpha +\beta_1 *\epsilon_t +\beta_2 * \epsilon_((t-1)) $$

```python
plot_acf(m);
plot_pacf(m);
```

```python
model=ARMA(m,order=(0,2))
result=model.fit()
```

```python
result.summary()
```

```python
plt.plot(result.resid)
```

```python
plot_acf((result.resid)**2);
plot_pacf((result.resid)**2);
```

### ARMA(1,1)

```python
ar=np.array([1,-0.7])
ma=np.array([1,0.2])
arma_process=ArmaProcess(ar,ma)
am=arma_process.generate_sample(nsample=1000)
```

```python
plt.plot(am)
```

```python
plot_acf(am,lags=100);
plot_pacf(am,lags=10);
```

$$ y_t = \alpha + \beta_1 y_ (t-1) +\beta_2 \epsilon_ (t-1) +\epsilon_t $$
where Epsilon at t-1 is a function of y at t-1

```python
model=ARMA(am,order=(2,2)).fit()
model.summary()
```

```python
model=ARMA(am,order=(4,1)).fit()
model.summary()
```

```python
model=ARMA(am,order=(1,1)).fit()
model.summary()
```
