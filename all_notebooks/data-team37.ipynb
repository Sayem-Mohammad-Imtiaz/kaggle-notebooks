```python
#นำเข้าข้อมูลโดยใช้คำสั่ง import 
import pandas as pd
import numpy as np
```

```python
#นำข้อมูลเข้าโดยใช้ไฟล์ tmdb_5000_movies.csv ไปที่ตัวแปล movies_df เพื่อนำไฟล์มาใช้งาน
movies_df = pd.read_csv('../input/tmdb-movie-metadata/tmdb_5000_movies.csv')
```

```python
#ทำการแสดงข้อมูลโดยประกอบด้วย งบประมาณ,ประเภท,เว็บที่อยู่,รหัส,ข้อความที่สำคัญ,ภาษา,ชื่อเรื่อง,ภาพรวม,คะแนนความนิยม,ชื่อบริษัท,ชื่อประเทศ,วัน/เดือน/ปี,รายได้,เวลา,ซับ,สถานะ,ลักษณะ,ชื่อเรื่อง,คะแนนเฉลี่ยของ Movie
movies_df.head()
```

**3.Wrangle, prepare, cleanse the data.**
     
     ทำการตรวจสอบข้อมูล และนำข้อมูลข้อมูลที่เราไม่ได้นำมาใช้ในการทำงานออกไป เพื่อให้ง่ายต่อการดู columns ที่จำเป็น

```python
#เราจะพบว่ามีข้อมูลหลายข้อมูลใน columns ที่เราไม่ได้นำมาใช้งาน
movies_df.columns
```

**4.Analyze, identify patterns, and explore the data.**
   
       ทำการสำรวจข้อมูล และวิเคราะห์ข้อมูล เปรียบเทียบว่าควรนำข้อมูลอะไรไปใช้ในการทำงาน เพื่อใช้ในการคำนวณ ซึ่งสิ่งที่จำเป็นต้องใช้มีดังนี้ คือ
            4.1 id
            4.2 vote_average
            4.3 vote_count
            4.4 popularity
            4.5 original_title

```python
#เลือกเฉพาะข้อมูลที่เราจะนำมาใช้งานซึ่งประกอบไปด้วย id,vote_average,vote_count,popularity,original_title
movies_clean = movies_df[[ 'id','vote_average','vote_count','popularity','original_title']]
#แสดงข้อมูลตามที่เราต้องการเพื่อนำไปใช้งาน
movies_clean.head()
```

```python
#เริ่มทำการกำหนดตัวแปลเพื่อใช้ในการคำนวณหาคะแนน 
R = movies_clean['vote_average']
v = movies_clean['vote_count']
m = movies_clean['vote_count'].quantile(0.7) ## setting threshold by selecting top 30% movies for this metric
C =  movies_clean['vote_average'].mean()
```

```python
#เราจะได้สูตรในการคำนวณ weighted_rating คือ ((R*v)+(C*m)) / (v+m)
movies_clean['weighted_rating'] = ((R*v)+(C*m)) / (v+m)
```

```python
#ทำการแสดงข้อมูลของ weighted_rating ที่ใช้สูตรในการคำนวณ
movies_clean
```

```python
# ทำการ import matplot กับ seaborn เพื่อมาใช้ในการแสดงกราฟ
# matplotlib.pyplot คือ module สำหรับวาดกราฟที่ใช้งานได้หลากหลาย
import matplotlib.pyplot as plt
# seaborn ใช้สำหรับการ plot ข้อมูลในรูปแบบต่าง ๆ
import seaborn as sns
#ทำการจัดเรียงข้อมูล Best average votes โดยการใช้ข้อมูลของ weighted_rating 
weight_average=movies_clean.sort_values('weighted_rating',ascending=False)
#เป็นการกำหนด ขนาดของกราฟโดยใช้ สูงกว้าง และใช้ facecolor ในการเปลี่ยนสีพื้นหลัง
plt.figure(figsize=(12,12),facecolor='#99DDFF')
#เป็นการกำหนดตำแหน่งของกราฟให้ title อยู่แกน x แล้วให้ rating อยู่แกน y โดยใช้ข้อมูลของ weight_average
axis1=sns.barplot(x=weight_average['original_title'].head(10), y=weight_average['weighted_rating'].head(10), data=weight_average)
#เป็นการกำหนดสัดส่วนของค่า x และ ค่า y
plt.xlim(-1, 9)
#เป็นการกำหนด ชื่อของกราฟ
plt.title('Best average votes', weight='bold')
#เป็นการกำหนดการใช้ชื่อตัวแปลของแกน x
plt.xlabel('Weighted Average Score', weight='bold')
#เป็นการกำหนดการใช้ชื่อตัวแปลของแกน Y
plt.ylabel('Movie Title', weight='bold');
#เป็นการกำหนดลักษณะของสีในแผนภูมิแท่งของกราฟ
plt.style.use('ggplot');
```

```python
#ทำการจัดเรียงข้อมูล Best Popularity Score โดยการใช้ข้อมูลของ weighted_rating 
weight_average_popularity=movies_clean.sort_values('popularity',ascending=False)
#เป็นการกำหนด ขนาดของกราฟโดยใช้ สูงกว้าง และใช้ facecolor ในการเปลี่ยนสีพื้นหลัง
plt.figure(figsize=(20,15),facecolor='#99DDFF')
#เป็นการกำหนดตำแหน่งของกราฟให้ title อยู่แกน x แล้วให้ popularity อยู่แกน y โดยใช้ข้อมูลของ weight_popularity
axis1=sns.barplot(x=weight_average_popularity['original_title'].head(10), y=weight_average_popularity['popularity'].head(10), data=weight_average_popularity);
#เป็นการกำหนดสัดส่วนของค่า x และ ค่า y
plt.xlim(-0.5, 9)
#เป็นการกำหนด ชื่อของกราฟ
plt.title('Best Popularity score', weight='bold')
#เป็นการกำหนดการใช้ชื่อตัวแปลของแกน x
plt.xlabel('Popularity Score', weight='bold')
#เป็นการกำหนดการใช้ชื่อตัวแปลของแกน Y
plt.ylabel('Movie Title', weight='bold');
#เป็นการกำหนดลักษณะของสีในแผนภูมิแท่งของกราฟ
plt.style.use('ggplot');
```

**5.Model, predict and solve the problem.**

     จากการวิเคราะห์ข้อมูล เราจะได้ข้อมูลมา 2 ชุด ซึ่งก้อคือ Average Score และ Popularity Score ซึ่งเราจะนำข้อมูลทั้ง 2 ชุดมา clustering เพื่อให้ได้ชุดข้อมูลที่เราต้องการ    

     

```python
#ใช้มอดูล sklearn เข้าช่วยในการจัดการเรื่องทำข้อมูลและ import MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
scaled_scores = pd.DataFrame(scalar.fit_transform(movies_clean[['popularity','weighted_rating']]), 
                             columns=['popularity_scaled','weighted_rating_scaled'])
movies_scaled = movies_clean.copy()
movies_scaled[['popularity_scaled','weighted_rating_scaled']] = scaled_scores
movies_scaled
```

```python
movies_scaled['score'] = (0.5 * movies_scaled['weighted_rating_scaled']) + \
                            (0.5 * movies_scaled['popularity_scaled'])
```

**6.Visualize, report, and present the problem solving steps and final solution.**

      จากการวิเคราะห์ข้อมูล ทำให้เราได้ข้อมูลมา 1 ชุด นั่นก้อคือ Best Movie by Hybrid Score ซึ่งจะมีข้อมูลเกี่ยวกับ Top 20 Movie ที่เราได้จัดอันดับโดยใช้ข้อมูลของ Average Score และ Popularity Score ซึ่งข้อมูลส่วนนี้สามารถนำไปใช้ในการตัดสิน ของผู้สร้างภาพยนตร์ หรือ ผู้กำกับภาพยนตร์ว่าจะสร้างหนังอะไร แนวไหน เพื่อที่จะให้ตอบโจทย์ความต้องการของผู้คนที่ชอบดูภาพยนตร์ อีกทั้งสามารถสร้างรายได้มหาศาล ถ้าเรารู้ว่าผู้คนชอบดูหนังแนวไหนโดยข้อมูลจะมีกราฟบอกอย่างชัดเจน 
      
      
      
      
   

```python
#ทำการจัดเรียงข้อมูล Best hybrid Score โดยการใช้ข้อมูลของ movies_scaled
weight_average_score_hybrid=movies_scaled.sort_values('score',ascending=False)
#เป็นการกำหนด ขนาดของกราฟโดยใช้ สูงกว้าง และใช้ facecolor ในการเปลี่ยนสีพื้นหลัง
plt.figure(figsize=(15,10),facecolor='#99DDFF')
#เป็นการกำหนดสัดส่วนของค่า x และ ค่า y
plt.xlim(-3, 20)
#เป็นการกำหนดตำแหน่งของกราฟให้ title อยู่แกน x แล้วให้ hybrid อยู่แกน y โดยใช้ข้อมูลของ weight_average_score_hybrid
axis1=sns.barplot(x=weight_average_score_hybrid['original_title'].head(20), y=weight_average_score_hybrid['score'].head(20), data=weight_average_score_hybrid);
#เป็นการกำหนด ชื่อของกราฟ
plt.title('Best Movies by Hybrid score', weight='bold')
#เป็นการกำหนดการใช้ชื่อตัวแปลของแกน x
plt.xlabel('Hybrid Score', weight='bold')
#เป็นการกำหนดการใช้ชื่อตัวแปลของแกน Y
plt.ylabel('Movie Title', weight='bold');
#เป็นการกำหนดลักษณะของสีในแผนภูมิแท่งของกราฟ
plt.style.use('ggplot');
```
