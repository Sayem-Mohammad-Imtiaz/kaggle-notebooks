## 3- Conditional Filtering

```python
#first condition: total bill>30
df[df['total_bill']>30]
```

```python
#second condition: sex=Male
df[df['sex']=='Male']
```

```python
#Multiple condition
df[(df['total_bill']>30) & (df['sex']=='Male')]
```

```python
df[(df['total_bill']>30) & (df['sex']!='Male')]
```

```python
#condition: Weekend
df[(df['day']=='Sun') | (df['day']=='Sat')]
```

```python
df[df['day'].isin(['Sun','Sat'])]
```
