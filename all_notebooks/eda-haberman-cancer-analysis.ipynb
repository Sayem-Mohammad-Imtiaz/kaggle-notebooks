# Univariate analysis

```python
#plotting histograms 1-D scatter plot
status_1=df.loc[df["status"]==1]
status_2=df.loc[df["status"]==2]

plt.plot(status_1["age"], np.zeros_like(status_1["age"]),"o",label="status_1")
plt.plot(status_2["age"],np.zeros_like(status_2["age"]),"o",label="status_2")
plt.title("AGE")
plt.grid()
plt.legend()

plt.show()
```

1]. Almost all points are overlapped<br>2]. Though, We can see patients whose age is less than 40 have lower rate of status_1 than whose age is above 40<br>3]. After 80 years we can only see status_2<br>4]. However patients whose age 75-78 have slightly status_1

```python
plt.plot(status_1["year"],np.zeros_like(status_1["year"]),"o",label="status_1")
plt.plot(status_2["year"],np.zeros_like(status_2["year"]),"o",label="status_2")
plt.grid()
plt.title("YEAR")
plt.legend()
plt.show()
```

1]. ALL points are overlapped and we can't have any insight from this

```python
plt.plot(status_1["nodes"],np.zeros_like(status_1["nodes"]),"o",label="status_1")
plt.plot(status_2["nodes"],np.zeros_like(status_2["nodes"]),"o",label="status_2")

plt.grid()
plt.legend()
plt.title("NODES")
plt.show()
```

1]. The points are overlapped very much.<br>2]. Though We can say that more patients have nodes between 0 to 25<br>3]. If we reverse status 1 and status 2 in this case, the also the points overlapped , so we can't see all points properly<br>4]. It is difficult or almost impossible to classify status_1 and status_2 using 1-D scatter plots

```python
#pdf of year
sn.FacetGrid(df,hue="status",height=4)\
  .map(sn.distplot, "year")\
  .add_legend()
plt.show()
```

1]. Two plots are overlapped very much, difficult to observe.

```python
#pdf of nodes
sn.FacetGrid(df,hue="status",height=4)\
  .map(sn.distplot, "nodes")\
  .add_legend()
plt.show()
```

1]. Though the plots are overlapped, It can be seen that patients with 0 nodes have higher rate of survival(status_1)<br>2]. patients with above 5 node have higher rate of status_2(death) than status _1(survival)<br>3]. both the plots are right skewed in nature

***. From these PDF plots also we can't classify status_1 and status_2 completely

# CDF

```python
# A]  plotting CDF
count , bin_edge = np.histogram(status_1["age"],bins=10,density=True)

pdf=count/sum(count)
print(pdf)
print(bin_edge)

cdf=np.cumsum(pdf)
plt.plot(bin_edge[1:],pdf,color="red")
plt.plot(bin_edge[1:],cdf,color="green")


count , bin_edge = np.histogram(status_2["age"],bins=10,density=True)

pdf=count/sum(count)
print(pdf)
print(bin_edge)

cdf=np.cumsum(pdf)
plt.plot(bin_edge[1:],pdf,color="blue")
plt.plot(bin_edge[1:],cdf,color="yellow")
plt.legend(["status_1_pdf","status_1_cdf","status_2_pdf","status_2_cdf"])
plt.grid()
plt.title("CDF for AGE")
plt.show()

```

1]. cdf curves are overlapped very much <br>2]. 20% of patients who survived(status_1) are of age below 40<br>3]. 20% of patients who do not survive(status_2) are of age approx. 45

```python
# B]
count , bin_edge = np.histogram(status_1["year"],bins=10,density=True)

pdf=count/sum(count)
print(pdf)
print(bin_edge)

cdf=np.cumsum(pdf)
plt.plot(bin_edge[1:],pdf,color="red")
plt.plot(bin_edge[1:],cdf,color="green")

count , bin_edge = np.histogram(status_2["year"],bins=10,density=True)

pdf=count/sum(count)
print(pdf)
print(bin_edge)

cdf=np.cumsum(pdf)
plt.plot(bin_edge[1:],pdf,color="blue")
plt.plot(bin_edge[1:],cdf,color="yellow")
plt.legend(["status_1_pdf","status_1_cdf","status_2_pdf","status_2_cdf"])
plt.title("CDF for YEAR")
plt.grid()
plt.show()
```

1]. CDF curves are highly overlapped , we can't distinguish bewtween status_1 and status_2


```python
#C]
count , bin_edge = np.histogram(status_1["nodes"],bins=10,density=True)

pdf=count/sum(count)
print(pdf)
print(bin_edge)

cdf=np.cumsum(pdf)
plt.plot(bin_edge[1:],pdf,color="red")
plt.plot(bin_edge[1:],cdf,color="green")

count , bin_edge = np.histogram(status_2["nodes"],bins=10,density=True)

pdf=count/sum(count)
print(pdf)
print(bin_edge)

cdf=np.cumsum(pdf)
plt.plot(bin_edge[1:],pdf,color="blue")
plt.plot(bin_edge[1:],cdf,color="yellow")
plt.grid()
plt.legend(["status_1_pdf","status_1_cdf","status_2_pdf","status_2_cdf"])
plt.title("CDF for NODES")
plt.show()

```

1]. though the curves are overlapped, we can see that, <br>* more than 85% of patients who survived(status_1) have nodes below 10.<br>* 80% of patients who do not survive(status_2) are having nodes more than 10


we should carryout bivariate analysis further

# Statistics(Mean,Median,IQR,Quantile,MAD)

```python
print("MEDIANS of age of status_1 and status_2 :\n")
print(np.median(status_1["age"]))
print(np.median(status_2["age"]))

print("\nMEDIANS of year of status_1 and status_2 :\n")
print(np.median(status_1["year"]))
print(np.median(status_2["year"]))

print("\nMEDIANS of nodes of status_1 and status_2 :\n")
print(np.median(status_1["nodes"]))
print(np.median(status_2["nodes"]))
```

```python
print("quantiles of age of status 1 and status2 :\n")
print(np.percentile(status_1["age"],np.arange(0,100,25)))
print(np.percentile(status_2["age"],np.arange(0,100,25)))

print("\nquantiles of year of status 1 and status2 :\n")
print(np.percentile(status_1["year"],np.arange(0,100,25)))
print(np.percentile(status_2["year"],np.arange(0,100,25)))

print("\nquantiles of nodes of status 1 and status2 :\n")
print(np.percentile(status_1["nodes"],np.arange(0,100,25)))
print(np.percentile(status_2["nodes"],np.arange(0,100,25)))
```

1].From quantiles of age,<br> for age below 34 there is no patients of status_2(died)<br>2]. From quantiles of nodes,<br> 75% of patients with status_1(survival) have nodes maximum of 3.  

```python
from statsmodels import robust
print("mean absolute median of age status_1 and status_2 : ")
print(robust.mad(status_1["age"]))
print(robust.mad(status_2["age"]))

print("\nmean absolute median of year status_1 and status_2 : ")
print(robust.mad(status_1["year"]))
print(robust.mad(status_2["year"]))

print("\nmean absolute median of nodes status_1 and status_2 : ")
print(robust.mad(status_1["nodes"]))
print(robust.mad(status_2["nodes"]))

```

# JOINT PLOTS

```python
sn.jointplot(x="age",y="year",data=status_1,kind="kde")

plt.title("status_1")

sn.jointplot(x="age",y="year",data=status_2,kind="kde")
plt.title("status_2")

plt.show()
```

***From the contour plot for status_1(survived):***<br>1]. More points are in the range of 1960-1962 and patients age approx 48-54.<br>***From the contour plot for status_2(died):***<br>1]. More points are at approx 1964 and patients age approx 50-51.

***From the contour plot for status_1(survived):***<br>1]. More points are in the range of nodes 0 and patients age approx 40-65.<br>***From the contour plot for status_2(died):***<br>1].
