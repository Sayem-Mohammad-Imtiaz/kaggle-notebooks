# Question 2

```python
odi['MatchDate'] = pd.to_datetime(odi['MatchDate'])
```

```python
def idealYears(data, player):
    
    playerCareer = data[data['Player'] == player]
    playerCareer['year'] = playerCareer['MatchDate'].dt.year
    allYears = set(playerCareer['year'])
    careerStart = playerCareer['year'].min()
    careerEnd = playerCareer['year'].max()
    
    count = 0
    for i in range(careerStart, careerEnd+1):
        if(i not in allYears):
            count = count + 1
            
    return count
```

```python
players = pd.DataFrame()
players['name'] = odi['Player'].unique()
players['idealYears'] = players['name'].apply(lambda x : idealYears(odi, x))
players = players.sort_values('idealYears', ascending = False)
top10Ideal = players.set_index('name')
top10Ideal = top10Ideal.head(10)
top10Ideal.plot.bar()
```

# Question 3

```python
def matchesTo2000(data, player):
    players = data[data['Player'] == player]
    players.reset_index(drop = True)
    if(players['Runs'].sum() < 2000):
        return 9999
    players = players.sort_values('MatchDate', ascending = False)
    players['cummulativeSum'] = players['Runs'].cumsum()
    subset = players[players['cummulativeSum'] <= 2000]
    return subset.shape[0]
```

```python
players = pd.DataFrame()
players['name'] = odi['Player'].unique()
players['matchesTo2000'] = players['name'].apply(lambda x : matchesTo2000(odi, x))
players = players.sort_values('matchesTo2000')
fastestTo2000 = players.set_index('name')
fastestTo2000 = fastestTo2000.head(10)
fastestTo2000.plot.bar()
```

# Question 4

```python
odi['is_century'] = 0
odi['is_century'][odi['Runs'] > 99] = 1
```

```python
def first10Centuries(data, player):
    players = data[data['Player'] == player]
    players.reset_index(drop = True)
    if(players['is_century'].sum() < 10):
        return 9999
    players = players.sort_values('MatchDate', ascending = False)
    players['totalCenturiesSofar'] = players['is_century'].cumsum()
    subset = players[players['totalCenturiesSofar'] <= 10]
    return subset.shape[0]
```

```python
players = pd.DataFrame()
players['name'] = odi['Player'].unique()
players['matchesTo10Centuries'] = players['name'].apply(lambda x : first10Centuries(odi, x))
players = players.sort_values('matchesTo10Centuries')
fastestToTen100s = players.set_index('name')
fastestToTen100s = fastestToTen100s.head(10)
fastestToTen100s.plot.bar()
```
