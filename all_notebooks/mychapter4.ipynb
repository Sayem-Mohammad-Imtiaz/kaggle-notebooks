```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import math
import random
def meanBootstrap(X, numberb):
    x = [0]* numberb
    for i in range (numberb):
        sample = [X[j]
            for j in np.random.randint(len(X), size=len(X))]
        x[i] = np.mean(sample)
    return x
m = meanBootstrap(accidents, 100)
print('Среднее распределение выборки путем начальной загрузки:', np.mean(m))
m = accidents.mean()
se = accidents.std() / math.sqrt(len(accidents))
ci = [m - se * 1.96, m + se * 1.96]
print('Доверительные интервалы:', ci)
```

```python
m = len(counts60)
n = len(counts30)
p = (counts30.mean() - counts60.mean())
print('m: ', m, 'n: ', n)
print('Средняя разница: ', p)
```

```python
y = counts60
x=counts30
pool = np.concatenate([x, y])
np.random.shuffle(pool)
N = 100 # number of samples
diff = list(range(N))
for i in range(N):
     p1 = [random.choice(pool) for _ in range(n)]
     p2 = [random.choice(pool) for _ in range(n)]
     diff[i] = (np.mean(p1) - np.mean(p2))
```

```python
diff2 = np.array(diff)
w1 = np.where(diff2 > p)[0]
print('p-value (Simulation)=', len(w1)/ float (N), '(', len(w1) / float(N) * 100 ,'%)', 'Difference =', p)
if (len(w1)/ float (N)) < 0.05:
     print('The effect is likely')
else:
     print('The effect is not likely')
```

```python
acc_time = data[['Date','Day_of_Week','Time']]
acc_time.head()
acc_time.dropna(axis=0,inplace=True)
def year(string):
    return int(string[6:10])
acc_time['Year']=acc_time['Date'].apply(lambda x: year(x))
def month(string):
    return int(string[3:5])
acc_time['Month']=acc_time['Date'].apply(lambda x: month(x))
def day(string):
    return int(string[0:2])
acc_time['Day']=acc_time['Date'].apply(lambda x: day(x))
def hour(string):
    s=string[0:2]
    return int(s)
acc_time['Hour']=acc_time['Time'].apply(lambda x: hour(x))

accidents = acc_time.groupby(['Hour']).size()
print(accidents.mean())
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
sns.set(style="darkgrid")
plt.figure(figsize=(15,8))
Hourplot = sns.countplot(x='Hour',data=acc_time)
Hourplot.set(xlabel='Hour', ylabel='Count')
for p in Hourplot.patches: 
    height = p.get_height() 
    Hourplot.text(p.get_x()+p.get_width()/2., 
      height + 3, 
      '{:1.2f}'.format((height/len(acc_time))*100)+'%',  
      ha="center") 
Hourplot.hlines(68352, 0, 24)
plt.show()

#график кол-ва происшествий по часам
```
