```python
data = pd.read_csv('/kaggle/input/data-science-job-posting-on-glassdoor/Uncleaned_DS_jobs.csv')
data.head()
```

```python
data.info()
```

```python
#Dropping index col.
data.drop(labels='index', axis=1,inplace=True)
data.head()
```

```python
data['Job Title'].value_counts()
```

```python
data['Job Title'].unique()
```

```python
for i in data['Job Title']:
    if "\n" in i:
        print (i) 
    else:
        continue
#There are no such values in Job Title Col. And the value seems        
```

```python
#Cleaning of values in Salary Est. Col so it can be used in further analysis.
data['Salary Estimate'] = data['Salary Estimate'].apply(lambda x: x.split('(') [0])
data['Salary Estimate'].unique()
```

```python
data['Salary Estimate'] = data['Salary Estimate'].apply(lambda x: x.replace('K', '').replace('$', ''))
data['Salary Estimate'].unique()
```

```python
#Solving \n values in Job Description.
data['Job Description'] = data['Job Description'].apply(lambda x: x.replace('\n', ' '))
data.head()
```

```python
data['Rating'].unique()
```

```python
data['Rating'] = np.where(data.Rating == -1.0, 0.0, data.Rating)
data['Rating'].unique()
```

```python
#Col-Company Name contains Name as well as Rating, solving it.
data['Company Name'] = data['Company Name'].apply(lambda x: x.split('\n') [0])
data['Company Name'].unique()
```

```python
#Adding a new col.
data['Same Location'] = data.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis = 1)
data.head()
```

```python
data['Founded'].unique()
data.Founded.dtype
```

```python
data.drop(labels = 'Competitors', axis = 1, inplace = True)
data.head()
```

```python
def python(skill):
    if 'python' in skill.lower():
        return 1
    else:
        return 0
```

```python
data['Python'] = data['Job Description'].apply(python)
```

```python
data.head()
```

```python
def excel(skill):
    if 'excel' in skill.lower():
        return 1
    else:
        return 0
def tableau(skill):
    if 'tableau' in skill.lower():
        return 1
    else:
        return 0
def power_bi(skill):
    if 'power bi' in skill.lower():
        return 1
    else:
        return 0
def big_data(skill):
    if 'big data' in skill.lower():
        return 1
    else:
        return 0
def hadoop(skill):
    if 'hadoop' in skill.lower():
        return 1
    else:
        return 0
# Alternative way.
data['AWS'] = data['Job Description'].apply(lambda x: 1 if 'aws' in x.lower() else 0)
data['Azure'] = data['Job Description'].apply(lambda x: 1 if 'azure' in x.lower() else 0)
data['Cloud'] = data['Job Description'].apply(lambda x: 1 if 'cloud' in x.lower() else 0)
```

```python
data['Excel'] = data['Job Description'].apply(excel)
data['Tableau'] = data['Job Description'].apply(tableau)
data['Power BI'] = data['Job Description'].apply(power_bi)
data['Big Data'] = data['Job Description'].apply(big_data)
data['Hadoop'] = data['Job Description'].apply(hadoop)
```

```python
data.head()
```

```python
data['Tableau'].unique()
```

```python
data['Power BI'].unique()
```

```python
def category(jt):
    if 'analyst' in jt.lower():
        return 'Analyst'
    elif 'data scientist' in jt.lower():
        return 'Data Scientist'
    elif 'machine learning' in jt.lower():
        return 'Machine Learning'
    elif 'data engineer' in jt.lower():
        return 'Data Engineer'
    elif 'manager' in jt.lower():
        return 'Manager'
    elif 'diretor' in jt.lower():
        return 'Director'
    else:
        return 'NA'
```

```python
data['Job Category'] = data['Job Title'].apply(category)
```

```python
data['Job Category'].value_counts()
```

```python
data.head()
```

```python
data.insert(2, 'MinSal', data['Salary Estimate'].apply(lambda x: x.split('-')[0]))
```

```python
data.insert(3, 'MaxSal', data['Salary Estimate'].apply(lambda x: x.split('-')[1]))
```

```python
data['MinSal'] = data['MinSal'].astype(int)
```

```python
data['MaxSal'] = data['MaxSal'].astype(int)
```

```python
data.insert(4, 'AvgSal', (data.MinSal+data.MaxSal)//2)
```
