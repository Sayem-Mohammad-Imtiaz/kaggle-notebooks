```python
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB
import pandas as pd
from sklearn.metrics import confusion_matrix
from sklearn import metrics
from sklearn.metrics import plot_confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_real = pd.read_csv('/kaggle/input/on-time-graduation-classification/data_lulus_tepat_waktu.csv')
data_train = pd.read_csv('/kaggle/input/dataalwi/datalulus.csv')
data_train.head()
```

```python
le = preprocessing.LabelEncoder()
data_train_df = pd.DataFrame(data_train)
data_train_df_encoded = data_train_df.apply(le.fit_transform)


data_real_df = pd.DataFrame(data_real)
data_real_df_encoded = data_real_df.apply(le.fit_transform)

x_train = data_train_df_encoded.drop(['tepat'],axis=1) 
y_train = data_train_df_encoded['tepat']

x_real = data_real_df_encoded.drop(['tepat'],axis=1)
y_real = data_real_df_encoded['tepat']

model = GaussianNB()
nbtrain = model.fit(x_train, y_train)

y_pred = nbtrain.predict(x_real)
print("Accuracy:",metrics.accuracy_score(y_real, y_pred))
```
