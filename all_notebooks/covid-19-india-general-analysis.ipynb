```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns  
from sklearn.model_selection  import train_test_split

covid_19_india = pd.read_csv('../input/covid19-in-india/covid_19_india.csv')
covid_19_india.head()


my_tab = pd.crosstab(index=covid_19_india["State/UnionTerritory"],  # Make a crosstab
                     columns="count")                  # Name the count column
my_tab

```

```python
covid_19_india.groupby(by=['State/UnionTerritory'])['Confirmed'].sum().reset_index().sort_values(['Confirmed']).tail(10).plot(x='State/UnionTerritory',
                                                                                                           y='Confirmed',
                                                                                                           kind='bar',
                                                                                                           figsize=(15,5))
plt.show()
```

```python
covid_19_india.groupby(by=['State/UnionTerritory'])['Cured'].sum().reset_index().sort_values(['Cured']).tail(10).plot(x='State/UnionTerritory',
                                                                                                           y='Cured',
                                                                                                           kind='bar',
                                                                                                           figsize=(15,5))
plt.show()
```

### Currently, in the above bar chart, we see Maharashtra has high number confirmed cases followed by Kerala.

```python
import plotly.express as px
cnf, dth, rec, act = '#393e46', '#ff2e63', '#21bf73', '#fe9801'
temp = covid_19_india.groupby('Date')['Confirmed', 'Deaths', 'Cured'].sum().reset_index()
temp = temp.melt(id_vars="Date", value_vars=['Confirmed', 'Deaths', 'Cured'],
                 var_name='Case', value_name='Count')
fig = px.area(temp, x="Date", y="Count", color='Case', height=600,
             title='Cases over time', color_discrete_sequence = [act, dth, rec])
fig.update_layout(xaxis_rangeslider_visible=True)
fig.show()
```

```python
import plotly.graph_objs as go
temp = covid_19_india.groupby('Date')['Confirmed', 'Deaths', 'Cured'].sum().reset_index()

fig = go.Figure(data=[
    go.Bar(name='Confirmed', x=temp["Date"], y=temp["Confirmed"], marker=go.bar.Marker(
    color=act)),
    go.Bar(name='Deaths', x=temp["Date"], y=temp["Deaths"],marker=go.bar.Marker(
    color=dth)),
    go.Bar(name='Cured', x=temp["Date"], y=temp["Cured"],marker=go.bar.Marker(
    color=rec))
])
# Change the bar mode
fig.update_layout(
    width = 900,
    height=500,
    barmode='stack', 
    title='Total cases over time',
    font=dict(size=14, color='#686868'),
    xaxis_tickangle=-45,
    yaxis=dict(
        title='Count'
        ))

fig.show()
```
