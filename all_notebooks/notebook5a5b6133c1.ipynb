```python
y= df['AbsentHours']
X= df.drop('AbsentHours',axis=1)
```

# 60:20:20 SPLIT

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1)

X_test, X_val, y_test, y_val = train_test_split(X_train, y_train, test_size=0.5, random_state=1) # 0.25 x 0.8 = 0.2
```

```python
print("Respective rows : ")
print("Training Data :",X_train.shape[0])
print("Validation Data :",X_val.shape[0])
print("Test Data :",X_test.shape[0])
```

# LINEAR REGRESSION MODEL

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_val_predicted = model.predict(X_val)
```

```python
from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(y_val, y_val_predicted)
 
RMSE = math.sqrt(MSE)
linearRMSE =RMSE
print(RMSE)
```

**Retrain on train and validation data and then compute RMSE for test data**

```python
model.fit(X_train, y_train)
model.fit(X_val, y_val)

y_test_predicted = model.predict(X_test)

MSE = mean_squared_error(y_test, y_test_predicted)
 
RMSE = math.sqrt(MSE)
print(RMSE)
```

# LASSO REGRESSION MODEL

```python
from sklearn.linear_model import Lasso
model = Lasso(alpha=1.0)

model.fit(X_train, y_train)
y_val_predicted = model.predict(X_val)
```

```python
from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(y_val, y_val_predicted)
 
RMSE = math.sqrt(MSE)
lassoRMSE =RMSE
print(RMSE)
```

**Retrain on train and validation data and then compute RMSE for test data**

```python
model.fit(X_train, y_train)
model.fit(X_val, y_val)

y_test_predicted = model.predict(X_test)

MSE = mean_squared_error(y_test, y_test_predicted)
 
RMSE = math.sqrt(MSE)
print(RMSE)
```

# RIDGE REGRESSION MODEL

```python
from sklearn.linear_model import Ridge
model = Ridge(alpha=1e-3,normalize=True)

model.fit(X_train, y_train)

y_val_predicted = model.predict(X_val)
```

```python
from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(y_val, y_val_predicted)
 
RMSE = math.sqrt(MSE)
ridgeRMSE = RMSE
print(RMSE)
```

**Retrain on train and validation data and then compute RMSE for test data**

```python
model.fit(X_train, y_train)
model.fit(X_val, y_val)

y_test_predicted = model.predict(X_test)

MSE = mean_squared_error(y_test, y_test_predicted)
 
RMSE = math.sqrt(MSE)
print(RMSE)
```

# The model with least Validation RMSE

```python
if linearRMSE < ridgeRMSE and linearRMSE < lassoRMSE:
    print("linear RMSE is the LEAST")
elif lassoRMSE < ridgeRMSE and lassoRMSE < linearRMSE:
    print("lasso RMSE is the LEAST")
else:
    print("Ridge RMSE is the LEAST")
```
