## This task requires you to create a ML model with the best score possible, from various models available.

```python
import pandas as pd
#mydata=pd.read_csv('UCI_Credit_Card.csv')
mydata=pd.read_csv('/kaggle/input/default-of-credit-card-clients-dataset/UCI_Credit_Card.csv')
mydata.shape
```

```python
import pandas as pd
import numpy as np
mydata['AGE'] = pd.cut(x=mydata['AGE'], bins=[10, 20, 30, 40, 50, 60, 70,80], labels=[1, 2,3,4,5,6,7])

mydata['AGE'].replace(to_replace=[1, 2, 3,4,5,6,7], value=[1, 2, 3,4,5,6,7], inplace=True)
```

```python

mydata.head(5)
```

```python
mydata.info()
```

```python
mydata = pd.get_dummies(mydata , columns= ['SEX','MARRIAGE'], drop_first = True)
mydata.columns
```

```python
mydata.head()
```

```python
mydata.dropna()
y = mydata[['default.payment.next.month']]
mydata.drop('default.payment.next.month', axis=1, inplace=True)
X = mydata.iloc[ : ,1:26]
print(X.columns)
X.shape
```

```python
X.info()
```

```python
X.head(5)
```

```python

from sklearn.preprocessing import RobustScaler
premodel=RobustScaler()
premodel.fit(X[['BILL_AMT1']]) 
X[['BILL_AMT1']] = premodel.transform(X[['BILL_AMT1']]) 
premodel.fit(X[['BILL_AMT2']]) 
X[['BILL_AMT2']] = premodel.transform(X[['BILL_AMT2']]) 
premodel.fit(X[['BILL_AMT3']]) 
X[['BILL_AMT3']] = premodel.transform(X[['BILL_AMT3']]) 
premodel.fit(X[['BILL_AMT4']]) 
X[['BILL_AMT4']] = premodel.transform(X[['BILL_AMT4']]) 
premodel.fit(X[['BILL_AMT5']]) 
X[['BILL_AMT5']] = premodel.transform(X[['BILL_AMT5']]) 
premodel.fit(X[['BILL_AMT6']]) 
X[['BILL_AMT6']] = premodel.transform(X[['BILL_AMT6']]) 

premodel.fit(X[['PAY_AMT1']]) 
X[['PAY_AMT1']] = premodel.transform(X[['PAY_AMT1']]) 
premodel.fit(X[['PAY_AMT2']]) 
X[['PAY_AMT2']] = premodel.transform(X[['PAY_AMT2']]) 
premodel.fit(X[['PAY_AMT3']]) 
X[['PAY_AMT3']] = premodel.transform(X[['PAY_AMT3']]) 
premodel.fit(X[['PAY_AMT4']]) 
X[['PAY_AMT4']] = premodel.transform(X[['PAY_AMT4']]) 
premodel.fit(X[['PAY_AMT5']]) 
X[['PAY_AMT5']] = premodel.transform(X[['PAY_AMT5']]) 
premodel.fit(X[['PAY_AMT6']]) 
X[['PAY_AMT6']] = premodel.transform(X[['PAY_AMT6']]) 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
from numpy.random import randint
from numpy.random import seed
from sklearn.metrics import classification_report, confusion_matrix
from scipy.special import expit
from sklearn.neighbors import KNeighborsClassifier
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, 
                                                    test_size = 0.2, random_state=190)

model1=KNeighborsClassifier(n_neighbors=2)
model1.fit(X_train,y_train)

print(model1.classes_)

print("Train score:")
print(model1.score(X_train, y_train))
print("Test score:")
print(model1.score(X_test, y_test))
y_predictions = model1.predict(X_test)
print(y_predictions)
print(y_predictions.sum())
print(y_train.sum())
print(confusion_matrix(y_test, y_predictions ))
print(classification_report(y_test,y_predictions ))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification


#X, y = make_classification(n_samples=1000, n_features=4 n_informative=2, n_redundant=0,random_state=0, shuffle=False)

#fit
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.8, 
                                                    test_size = 0.2, random_state=108)
model2 = RandomForestClassifier(n_estimators = 100, 
                                    max_depth=50,
                                    random_state = 108)
y_predictions = model2.fit(X_train, y_train).predict(X_test)
print("Train score:")
print(model2.score(X_train, y_train))
print("Test score:")
print(model2.score(X_test, y_test))

print(y_predictions)
print(y_predictions.sum())
print(y_train.sum())
```
