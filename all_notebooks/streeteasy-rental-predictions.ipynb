```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Let’s take a look at the first few rows

```python
df = pd.read_csv('../input/streeteasy-dataset/manhattan.csv')
df.head()
```

### Check missing values

```python
df.isnull().sum()
```

```python
# summarize the data
df.describe()
```

```python
x = df[['bedrooms', 'bathrooms', 'size_sqft', 'min_to_subway', 'floor', 'building_age_yrs', 'no_fee', 'has_roofdeck', 'has_washer_dryer', 'has_doorman', 'has_elevator', 'has_dishwasher', 'has_patio', 'has_gym']]
y = df[['rent']]
```

### Split dataset into training set and test sets

Use scikit-learn’s train_test_split() method to split x into 80% training set and 20%.

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size= 0.8, random_state=42)
```

### Let’s take a look at the shapes of `x_train`, `x_test`, `y_train`, and `y_test`

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

### Multiple Linear Regression: Scikit-Learn

Now we have the training set and the test set, let’s use scikit-learn to build the linear regression model.

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
# find the coefficient and the intercept
model.fit(x_train, y_train)

# predict y-values from x_test
y_predict = model.predict(x_test)
```

### Let's test it

```python
john_appartment = [[1, 1, 620, 16, 1, 98, 1, 0, 1, 0, 0, 1, 1, 0]]
yy = model.predict(john_appartment)
print(yy)
```

### Visualizing Results with Matplotlib

Create a 2D scatter plot using `y_test` and `y_predict`.

```python
plt.scatter(y_test,y_predict,alpha=0.4)
plt.xlabel('t_test')
plt.ylabel('Y_predict')
plt.title('Graph')
plt.show()
```

### Evaluating the Model's Accuracy

```python
print('Train score:')
print(model.score(x_train, y_train))
print('Test score:')
print(model.score(x_test, y_test))
```

### Graph a scatter plot of residuals vs. predicted_y values:
we can use **Residual Analysis**  to evaluate the accuracy of our multiple linear regression model.

```python
residuals = y_predict - y_test
 
plt.scatter(y_predict, residuals, alpha=0.4)
plt.title('Residual Analysis')
 
plt.show()
```

**In the end** We can Remove some of the features that don’t have strong correlations and see if our scores improved!

```python
plt.scatter(y_test, y_predict)
plt.plot(range(20000), range(20000))

plt.xlabel("Prices: $Y_i$")
plt.ylabel("Predicted prices: $\hat{Y}_i$")
plt.title("Actual Rent vs Predicted Rent")

plt.show()
```
