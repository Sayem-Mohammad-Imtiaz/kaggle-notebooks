```python
def trans(trans):
    if trans == 'Automatic':
        return 'Automatic'
    elif trans == 'Manual':
        return 'Manual'
    else:
        return np.nan
```

```python
data['transmission'] = data['transmission'].apply(trans)
```

```python
cars = data.drop(['id','model','title','engine'],axis=1)
```

```python
cars['body'].value_counts()
```

```python
body_dum = pd.get_dummies(cars['body'],drop_first=True
                         )
```

```python
cars = pd.concat([cars,body_dum,discount_dum,brand_dum,trans_dum,seller_dum,state_dum],axis=1)
```

```python
x = cars.drop(['body','discount','brand','transmission','state','seller','year','price'],axis=1)
```

```python
y = cars['price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python

from sklearn.ensemble import RandomForestRegressor
```

```python
f_model = RandomForestRegressor(n_estimators=500)
```

```python
f_model.fit(x_train,y_train)
```

```python
predictions = f_model.predict(x_test)
```

```python
sns.scatterplot(y_test,predictions)
plt.plot(y_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mean_absolute_error(y_test,predictions)
```

```python
sns.distplot(predictions - y_test)
```
