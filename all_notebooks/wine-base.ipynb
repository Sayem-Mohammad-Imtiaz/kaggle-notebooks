```python
import pandas as pd
df = pd.read_csv('/kaggle/input/wine-information/wine_info.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
df['year'] = df['year'].fillna(0)
df['price'] = df['price'].fillna(0)
df['ml'] = df['ml'].fillna(0)
```

```python
df.isna().sum()
```

```python
df['year'] = df['year'].astype('int64')
df['price'] = df['price'].astype('int64')
df['ml'] = df['ml'].astype('int64')
```

```python
import matplotlib.pyplot as plt

plt.plot(df.price)
plt.show()
```

```python
import re
pd.set_option('mode.chained_assignment',  None)

for i in range(len(df)):
    df['producer'][i] = re.sub('[^ a-zA-Z0-9]','',df['producer'][i]).strip()
    df['nation'][i] = re.sub('[^ a-zA-Z0-9]','',df['nation'][i]).strip()
    try:
        df['local1'][i] = re.sub('[^ a-zA-Z0-9]','',df['local1'][i]).strip()
        df['local2'][i] = re.sub('[^ a-zA-Z0-9]','',df['local2'][i]).strip()
        df['local3'][i] = re.sub('[^ a-zA-Z0-9]','',df['local3'][i]).strip()
        df['local4'][i] = re.sub('[^ a-zA-Z0-9]','',df['local4'][i]).strip()
    except:
        continue
df
```

```python
plt.figure(figsize=(12,12))
plt.pie(df.nation.value_counts(), labels=df.nation.value_counts().index, autopct='%.2f%%')
plt.show()
```

```python
plt.figure(figsize=(25,5))
plt.xticks(rotation=45)
plt.plot(df.nation.value_counts())
plt.show()
```
