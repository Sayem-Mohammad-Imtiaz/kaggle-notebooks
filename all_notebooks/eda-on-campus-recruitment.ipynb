## First of all Let's check the missing values

```python
df.isna().sum()
```

## As we can see the Some of the data is missing from the column salary
      - we have two ways Eithe Delete the whole Column or Impute the missing Value

## Lets impute Missing values . Becoz it's the important column
      - We can get so many insights from that column.

```python
df["salary"].mean()
```

```python
df["salary"].mode()
```

```python
df["salary"].median()
```

```python
df["salary"]=df["salary"].fillna(df["salary"].mean())
df["salary"]=df["salary"].apply(np.round)
```

```python
df.head()
```

## Let's Do Some Analysis

## 01. Gender

```python
df["gender"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.title("% OF MALE AND FEMALE")
plt.legend()
plt.show()
```

```python
pd.crosstab(df["gender"],df["status"]).plot(kind="bar")
plt.title("HOW MANY MALE AND FEMALE GOT PLACED")

plt.show()
```

```python
df.groupby(by=df["gender"])["salary"].mean().plot(kind="bar")
plt.title("AVERAGE SALARY OF MALE AND FEMALE")
plt.show()
```

 ## Conclusion based on Gender.
 ###  - 01. 64.7% are Male & 35.3% are Female
 ###  - 02. As we can see Most MALES were placed as compared to FEMALES.
 ###  - 03.Now SALARY. MALES got little bit high salary as compared to FEMALES.

## 02. STATUS

```python
df.status.value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.title("%AGE OF OVERALL PLACEMENT")
plt.legend()
plt.show()
```

```python
pd.crosstab(df["status"],df["gender"]).mean().plot(kind="pie",autopct="%1.1f%%")
plt.title("%AGE OF PLACEMENT BASED ON GENDER")
plt.show()
```

```python
df["mba_p"]=df["mba_p"].apply(np.round)
```

```python
sns.boxplot(x="status",y="mba_p",data=df)
plt.title("PLACEMENT BASED ON MBA%AGE")
plt.show()
```

```python
print("BASED ON %AGE IN MBA. NUMBER OF STUDENTS GOT PLACED")
df.groupby(df["mba_p"])["status"].value_counts().sort_values(ascending=False)


```

```python
sns.boxplot(x="status",y="etest_p",data=df)
plt.title("PLACEMENT BASED ON  EMPLOY-TEST")
plt.show()
```

```python
sns.boxplot(x="status",y="degree_p",data=df)
plt.title("PLACEMENT BASED ON DEGREE %AGE")
plt.show()
```

```python
sns.boxplot(x="status",y="hsc_p",data=df)
plt.title("PLACEMENT BASED ON HSC %AGE")
plt.show()
```

```python
sns.boxplot(x="status",y="ssc_p",data=df)
plt.title("PLACEMENT BASED ON SSC %AGE")
plt.show()
```

```python
pd.crosstab(df["specialisation"],df["status"]).plot(kind="bar")
plt.title("PLACEMENT BASED ON SPECIALISATION ")
plt.show()
```

## Conclusion based on Status.
### - 01. Overall %age of Placed is 68.8%
### - 02. 64.7% MALES got placed & 35.3% FEMALES got placed.
### - 03. Based on Score in MBA .It doesn't  really affect the placement.AS we can see MOst of ones who got placed they got average %age in MBA . There are some with high %AGE still not got placed.
### - 04. Employability test percentage ( conducted by college) It also doesn't really affect the placemnt of students.
### - 05. There is Good relation between DEGREE% and placement. One who got high %age in DEGREE have high chances to get placed.
### - 06. HSC%AGE also affect the overall placement. One who got high %AGE in HSC have high chances to get placed.
### - 07. SSC%AGE also have a good Affect over placement. One who got high %AGE In SSC have high chances to get placed.
### - 08. There is also a good relation between SPECIALISATION and PLACEMENT.There is high chances to get placed if you have done MARKETING & FINANCE as compared to MARKETING & HR.
