# **Data and Library Import**

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score

vw_df = pd.read_csv("../input/used-car-dataset-ford-and-mercedes/vw.csv")
```

# **Data Exploration**

```python
print(vw_df.shape)
vw_df.head()
```

```python
print(vw_df.info())
print(vw_df.describe())
```

```python
# Palette
palette = sns.color_palette()

# Creating CountPlot for Car Model
fig, ax = plt.subplots(figsize = (10,5))
sns.countplot(y = "model", data = vw_df, order = vw_df['model'].value_counts().index)
plt.ylabel("VW Model")
plt.title("Distribution of VW Model")
```

```python
# Creating plot for

# Create subplot and define spacing
fig, axes = plt.subplots(2, figsize = (10,10))
fig.tight_layout(pad = 5)

# Subplot 1
sns.countplot(x = 'transmission', data = vw_df, ax = axes[0])
axes[0].set_title('Car Transmission Distribution')
axes[0].set_xlabel('Transmission')
axes[0].set_ylabel('Number of Cars')

# Subplot 2
sns.countplot(x = 'fuelType', data = vw_df, ax = axes[1])
axes[1].set_title('Car Fuel Type Distribution')
axes[1].set_xlabel('Fuel Type')
axes[1].set_ylabel('Number of Cars')
```

**The graph below shows that price varies more in older car**

```python
fig, ax = plt.subplots(figsize = (12,5))
sns.barplot(x = 'year', y = 'price', data = vw_df)
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = 'mileage', y = 'price', data = vw_df, hue = vw_df['year'])
plt.title('Mileage vs Price')
```
