```python
covid_data['cdc_report_dt'] = pd.to_datetime(covid_data['cdc_report_dt'])
```

```python
covid_data.groupby("cdc_report_dt")['sex'].value_counts()
```

```python
covid_data['cdc_report_day'] = covid_data['cdc_report_dt'].dt.day
covid_data['cdc_report_month'] = covid_data['cdc_report_dt'].dt.month
covid_data['cdc_report_day_name'] = covid_data['cdc_report_dt'].dt.day_name()
```

```python
sns.countplot(x='cdc_report_day_name',data=covid_data)
```

```python
sns.countplot(x='cdc_report_month',data=covid_data)
```

```python
sns.countplot(x='cdc_report_day',data=covid_data)
```

```python
covid_data['cdc_report_quater'] = covid_data['cdc_report_dt'].dt.quarter
sns.countplot(x='cdc_report_quater',data=covid_data)
```

```python
percent_missing = covid_data.isnull().sum() * 100 / len(covid_data)
missing_value_df = pd.DataFrame({'column_name': covid_data.columns,
                                 'percent_missing': percent_missing})
```
