```python
profiles_df = pd.read_csv("/kaggle/input/darts-players-top-500/PlayerProfiles.csv")
top_df = pd.read_csv("/kaggle/input/darts-players-top-500/Top500DartsPlayers.csv")
```

```python
profiles_df = profiles_df.rename(columns={'Name :': 'name' \
                                          , 'Country :': 'country' \
                                          , 'Age :': 'age' \
                                          , 'Date Of Birth :': 'dof' \
                                          , 'Nickname :': 'nick' \
                                          , 'PDC Ranking :': 'PDC' \
                                          , 'Tour Card :': 'tour_card' \
                                          , 'Career Earnings :': 'total_earnings'})
```

```python
top_df = top_df.rename(columns=lambda x: x.lower())
```

```python
profiles_df['total_earnings'] = profiles_df['total_earnings'].str.replace(",", "").str.replace("£", "").astype(int)
```

```python
profiles_df
```

<font size = 5 >I had some duplicates so lets drop them</font>

```python
profiles_df = profiles_df.drop_duplicates(subset='name')
```

```python
top_df['prize'] = top_df['prize'].str.replace('£', '').str.replace(',', '').astype(int)
```

<font size=8>Lets see what country is the best.</font>

```python
country_df = profiles_df.groupby('country', as_index=False).agg({'PDC': 'mean', 'name': 'count','total_earnings': 'sum'}) \
    .rename(columns={'name': 'total_players'})
```

```python
country_df
```

<font size= 6> This is temp dataframes to make vizualization easier</font>

```python
sorted_countries = country_df.sort_values('total_earnings', ascending=False)
```

```python
players_sorted = country_df.sort_values('total_players', ascending=False)
```

```python
sorted_pdc = country_df.sort_values('PDC', ascending=True)
```

<font size = 6> Here is bar plot for avarage PDC places(Less is better)</font>

```python
plt.figure(figsize=(20, 14))
ax = sns.barplot(y=sorted_pdc.PDC, x=sorted_pdc.country)
```

<font size = 6> So avg PDC highest in scotland, but lets see how many players was playing in each country</font>

```python
plt.figure(figsize=(20, 14))
sns.barplot(y=players_sorted.total_players, x=players_sorted.country)
```

<font size = 6> </font>

<font size = 6> Here is the total ammount of money made by players form each country</font>

```python
plt.figure(figsize=(20, 14))
sns.barplot(y=sorted_countries.total_earnings, x=sorted_countries.country)
```

<font size = 6> But lets create metric which will show avg income of each player </font>

```python
country_df['avg_player_income'] = round(country_df['total_earnings'] / country_df['total_players'], 2)
```

```python
income_sorted = country_df.sort_values('avg_player_income', ascending=False)
```

<font size = 6> Here  we can see that richest players are in scotland, there are not too many players, but their PDC is the highest</font>

```python
plt.figure(figsize=(20, 14))
sns.barplot(y=income_sorted.avg_player_income, x=income_sorted.country)
```

<font size = 6> Also I want to see total earnings of players from their age</font>

```python
ages_df = profiles_df.groupby('age', as_index=False).agg({'total_earnings': 'sum'})
```

```python
sns.relplot(data=ages_df, y='total_earnings', x='age', kind='line')
```
