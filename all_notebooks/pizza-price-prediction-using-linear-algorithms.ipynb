```python
#import required Lib.............
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split as tts 
```

```python
df = pd.read_csv('../input/pizza-price-prediction/pizza_v1.csv')# Read file
display(df.head(129))#View file..
```

```python
df.info() #information about the dataset
```

```python
df.isnull().sum() #Chech any missing value in dataset....
```

```python
#remove String & special character in price column..... && convert into 'int64'
df['price_rupiah'] = df['price_rupiah'].str.replace('Rp','').str.replace(',','')
df['price_rupiah'] = pd.to_numeric(df['price_rupiah'])
```

```python
#Check the data type of price column -//- is numeric...
df['price_rupiah'].dtype
```

```python
#Encode Label's.. in Dataset.Using SKlearn >>
EN = LabelEncoder()
df['company']= EN.fit_transform(df['company'])
df['topping']= EN.fit_transform(df['topping'])
df['variant']= EN.fit_transform(df['variant'])
df['size']= EN.fit_transform(df['size'])
df['extra_sauce']= EN.fit_transform(df['extra_sauce'])
df['extra_cheese']= EN.fit_transform(df['extra_cheese'])
df
```

```python
df.info() #once again check the dataset after the Encoder process....
```

```python
sns.heatmap(df.corr(),cmap= 'prism',annot = True) #co-relation chart....
```

```python
sns.scatterplot(df['diameter'], df['price_rupiah']) #visuval the data
```

```python
sns.scatterplot(df['company'], df['price_rupiah'])
```

```python
sns.scatterplot(df['topping'], df['price_rupiah'])
```

```python
#Split the Data to train and test the model....
x = df['diameter']
y = df['price_rupiah']
train_x,test_x,train_y,test_y = tts(x,y)
```

```python
lr = LinearRegression()
lr.fit(np.array(train_x).reshape(-1, 1), train_y)
```

```python
plt.scatter(x, y)
plt.plot(train_x, lr.coef_*train_x + lr.intercept_, '-r')
```

```python
predict_y = lr.predict(np.array(test_x).reshape(-1,1)) 
predict_y
```

```python
plt.scatter(test_y,predict_y)
plt.title("Actual & Predict Value")
plt.xlabel("Actual Value")
plt.ylabel("Predict Value")
plt.xlim(0, 250000)
plt.ylim(0, 250000)
plt.grid(True)
plt.show()
```

```python
#Save the Predict price amount value...
output = pd.DataFrame({'Actual price':test_y,'Predict price':predict_y})
output.to_csv('pizza price pred.csv', index =False)
output.head(5)
```
