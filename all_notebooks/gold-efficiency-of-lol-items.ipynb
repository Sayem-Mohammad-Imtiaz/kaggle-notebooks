## Part 1 : Stats/Gold Ratio

The first thing to do to go deeper in Gold efficiency is to make different assumptions.

Let's call 1 point of statistic every unity of statistic column. This means that 1 HP = 1 Stat point but also that 1% of critic hit = 1 Stat point.

To make this assumption true, we need to turn our decimals in percentages.

```python
dataitem[['AS']]=dataitem[['AS']]*100
```

```python
dataitem[['Crit','LS','APen','MP5','HSP','OVamp','MPen','HP5']]=dataitem[['Crit','LS','APen','MP5','HSP','OVamp','MPen','HP5']]*100
```

```python
dataitem
```

```python
dataitemstat = dataitem[['AD','AS','Crit','LS','APen','AP','AH','Mana','MP5','HSP','OVamp','MPen','Health','Armor','MR','HP5','MS']]
dataitemstat
```

#### Now we have set our data on an equal footing.

It is now time to structure the first part of our study : to calculate the Stat/ Gold Ratio.

```python
itemcost = dataitem[['Item','Cost']]
itemcost
```

```python
sumstat=dataitemstat.sum(axis=1)
sumstat
```

```python
goldstat = pd.concat([itemcost,sumstat], axis=1)
```

```python
goldstat
```

```python
goldstat=goldstat.rename(columns = {0: 'Sumstat'})
```

```python
goldstat
```

```python
ratiocal = pd.DataFrame((goldstat.Sumstat/goldstat.Cost)*100)
StatGoldRatio = pd.concat([goldstat,ratiocal], axis=1)
StatGoldRatio = StatGoldRatio.sort_values(by=0)
StatGoldRatio=StatGoldRatio.rename(columns = {0: 'Stats/Gold ratio'})
StatGoldRatio
```

#### Here we have our Stats/Gold ratio.
This allows us to realize how strong can be the raw statistics of the objects. But this ratio is not the gold efficiency we are looking for.
