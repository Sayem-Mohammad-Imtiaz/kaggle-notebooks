```python
df=pd.read_csv('/kaggle/input/titanic/train.csv')
```

```python
df.info()
```

```python
df.describe()
```

```python
df_missing_values=df.isnull().sum()
df_missing_values
```

```python
#categorizing into features and lables
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
r=['PassengerId','Name','Ticket','Cabin']
df1=df.drop(r,axis=1).copy()
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1['Embarked'].mode()[0]
```

```python
df1['Embarked']=df1['Embarked'].fillna(df1['Embarked'].mode()[0])
```

```python
df1['Age']=df1['Age'].fillna(df1['Age'].mean())
```

```python
df1.head()
```

```python
cc=['Sex','Embarked']
```

```python
dummy=pd.get_dummies(df1[cc])
dummy.head()
```

```python
df1=df1.drop(cc,axis=1)
```

```python
df1.head()
```

```python
df2=pd.concat([df1,dummy],axis=1)
df2.head()
```

```python
X=df2.iloc[:,1:]
X
```

```python
y=df2.iloc[:,0]
y
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y,test_size=0.25)
```

```python
# importing voting classifer
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import log_loss
```

```python
from sklearn import tree
# Building Decision Tree model 
dtc = tree.DecisionTreeClassifier(random_state=0)
dtc.fit(X_train, y_train)
```

```python
dtc.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=30)
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test)
```

```python
from sklearn.naive_bayes  import GaussianNB
nb=GaussianNB()
nb.fit(X_train,y_train)
```

```python
nb.score(X_test,y_test)
```

```python
model_1=tree.DecisionTreeClassifier(random_state=0)
model_2=RandomForestClassifier(n_estimators=30)
model_3=GaussianNB()
```

```python
final_model = VotingClassifier(
    estimators=[('dtc', model_1), ('clf', model_2), ('nb', model_3)], voting='hard')
```

```python
# training all the model on the train dataset
final_model.fit(X_train, y_train)
```

```python
# predicting the output on the test dataset
pred_final = final_model.predict(X_test)
pred_final
```

```python
# printing log loss between actual and predicted value
print(log_loss(y_test, pred_final))
```
