```python
data = pd.read_csv("/kaggle/input/car-evaluation-data-set/car_evaluation.csv")
```

```python
data.head()
```

```python
data.columns = ["buying", "maintenance", "no_doors", "no_persons", "luggage_boot", "safety", "class"]
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
X=data.drop(['class'],axis=1)
```

```python
y=data.filter(['class'])
```

```python
X.head()
```

```python
y.head()
```

```python
map = {
    "class":{
        "unacc":4,
        "acc":3,
        "good":2,
        "vgood":1
    }
}
```

```python
y=y.replace(map)
```

```python
y.head()
```

```python
X['no_doors'].value_counts()
```

```python
X['no_persons'].value_counts()
```

```python
X = pd.get_dummies(X)
```

```python
X.head()
```

```python
def correlation(hemanth,thresold):
    data=hemanth.corr()
    corr_features= set()
    for i in range(len(data.columns)):
        for j in range(i):
            if(abs(data.iloc[i,j])>thresold):
                colname=data.columns[i]
                corr_features.add(colname) 
    return corr_features
```

```python
features= correlation(X,0.7)
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)
```

```python
 y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,y_pred)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
tree.plot_tree(clf,filled=True)
```
