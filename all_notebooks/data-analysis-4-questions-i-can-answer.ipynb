### Question 3: Genre preference change during 10 years?

```python
# Preference in 2009 (Fiction vs non-Fiction)
fiction = len(data[(data["Year"] == 2009)& (data["Genre"] == "Fiction")])
non_fiction = len(data[(data["Year"] == 2009)& (data["Genre"] == "Non Fiction")])
labels = ["Fiction", "Non Fiction"]
size = [fiction, non_fiction]
explode = (0,0)
colors = ["#df7861", "#ecb390"]
fig1, ax1 = plt.subplots()
ax1.pie(size, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90, colors = colors)
ax1.axis('equal')  

plt.show()
```

```python
# Preference in 2019 (Fiction vs non-Fiction)
fiction = len(data[(data["Year"] == 2019)& (data["Genre"] == "Fiction")])
non_fiction = len(data[(data["Year"] == 2019)& (data["Genre"] == "Non Fiction")])
labels = ["Fiction", "Non Fiction"]
size = [fiction, non_fiction]
explode = (0,0)
colors = ["#df7861", "#ecb390"]
fig1, ax1 = plt.subplots()
ax1.pie(size, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90, colors = colors)
ax1.axis('equal')  

plt.show()
```

From 2009 to 2019, the preference of fiction books has reduced from 48% to 40%, the change is slight, so we can not make sure that the preference has been obviously changed.
Therefore, we need to generate the preference of each year to determine that:

```python
fiction_year = []
for year in range(2009, 2020):
    fic = len(data[(data["Year"] == year)& (data["Genre"] == "Fiction")])
    fiction_year.append(fic)
x = [i for i in range(2009, 2020)]
plt.plot(x, fiction_year)
```

The line chart shows the changes of preferences for fiction books, we can not see any trends or tendency of preferences changes, thus we can roughly make sure the perferences of audience does not have obvious changes during 2009 to 2019.
