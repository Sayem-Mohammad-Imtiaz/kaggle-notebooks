```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv("../input/creditcard/creditcard.csv")
```

```python
dataset.head(10)
```

```python
dataset.isnull().sum()
```

```python
#distributuion of legit and fradulent data
#0 -> normal transaction
#1 -> fradulent transaction
dataset['Class'].value_counts()
```

```python
#seperating the data for  analysis
legit = dataset[dataset.Class==0]
fraud = dataset[dataset.Class==1]
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
dataset.groupby('Class').mean()
```

```python
#UnderSampling
#Build a Sample containing similar distribution of normal  transaction  and Fradulent transaction
#Number of fradulent transaction -> 492
legit_sample = legit.sample(n=492)
```

```python
#Concatenate two dataset
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.groupby('Class').mean()
```

```python
#Splitting the data into feature and Targets
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting data into Training and test dataset
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=.2,stratify=Y,random_state=2)
```

```python
#Model Training
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
traning_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print(traning_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print(test_data_accuracy)
```
