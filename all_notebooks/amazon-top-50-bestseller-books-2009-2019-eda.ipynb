**Getting extra information for User rating**

```python
popularity = []
for i in data["User Rating"]:
    if i >= 4.8:
        popularity.append('Extremely Popular')
    elif i >= 4.5:
        popularity.append('Very Popular')
    elif i >= 4.0:
        popularity.append('Fairly Popular')
    else:
        popularity.append('Popular')
        
popularity = pd.DataFrame(popularity)
data = pd.concat([data,popularity],axis=1)
data.rename(columns = {0:'Popularity'}, inplace= True)
```

```python
data.head()
```

*Here we get some insight for User Rating called Popularity*
1. User rating higher than 4.7 is interpreted as Extremely Popular
2. User rating higher than 4.5 and less than 4.7 is interpreted as Very Popular
3. User rating higher than 4.0 and less than 4.5 is interpreted as Fairly Popular
4. User rating below than 4.0 is interpreted as Popular
