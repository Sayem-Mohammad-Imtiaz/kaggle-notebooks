```python
import pandas as pd
df=pd.read_csv("/kaggle/input/glass/glass.csv")

df.head()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x.head()
```

```python
y.value_counts()
```

```python
y.unique()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,  stratify=y,  test_size=0.25)
```

```python
from sklearn import tree

# Building Decision Tree model 
dtc = tree.DecisionTreeClassifier(random_state=0)
dtc.fit(x_train, y_train)
```

```python
dtc.score(x_test,y_test)
```

```python
from sklearn.model_selection import train_test_split
sn=[]
score=[]
for i in range(1,100):
    x_train, x_test, y_train, y_test = train_test_split(x, y,  stratify=y,  test_size=0.25)  # split
    dtc = tree.DecisionTreeClassifier(random_state=0)# model selection
    dtc.fit(x_train, y_train)# model learning
    sn.append(i)
    score.append(dtc.score(x_test,y_test))

result=pd.DataFrame()
result["SN"]=sn
result["Score"]=score
```

```python
result
```

```python
result.describe()
```

```python
result.to_csv("r1.csv")
```

```python
from sklearn import preprocessing

minmax_scale=preprocessing.MinMaxScaler(feature_range=(0,10))

x_minmax=minmax_scale.fit_transform(x)
```

```python
x_minmax
```

```python
x1=pd.DataFrame(columns=x.columns,data=x_minmax)

x1
```

```python
sn=[]
score=[]
for i in range(1,100):
    from sklearn.model_selection import train_test_split
    x_train, x_test, y_train, y_test = train_test_split(x1, y,  stratify=y,  test_size=0.25)
    dtc = tree.DecisionTreeClassifier(random_state=0)
    dtc.fit(x_train, y_train)
    sn.append(i)
    score.append(dtc.score(x_test,y_test))

result=pd.DataFrame()
result["SN"]=sn
result["Score"]=score
print(result)    
```
