```python
X= df.iloc[:,1:]
y= df.iloc[:,:1]
```

```python
from sklearn.decomposition import PCA
pca= PCA(n_components=int(len(cols)/4))
X= pca.fit_transform(X.values)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.2)
```

```python
import warnings
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

clf= XGBClassifier(random_state=42)

with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    model = clf.fit(X_train, y_train)
    y_preds= clf.predict(X_test)
    acc= accuracy_score(y_preds, y_test)


print(f'The accuracy score is: {acc*100:.3f}%')
```

```python
from sklearn.ensemble import RandomForestClassifier
clf= RandomForestClassifier(random_state=42)
model= clf.fit(X_train, y_train)
y_preds= model.predict(X_test)
acc= accuracy_score(y_preds, y_test)

print(f'The accuracy score is: {acc*100:.3f}%')
```

```python
from sklearn.metrics import f1_score, precision_score, recall_score
f1= f1_score(y_preds, y_test)
prec= precision_score(y_preds, y_test)
recall= recall_score(y_preds, y_test)

print(f'The precision score is: {prec}')
print(f'The recall score is: {recall}')
print(f'The f1 score is: {f1}')
```

## Conclusion
After getting 96.041% and 96.114% from XGBClassifier and RandomForestClassifier respectivly, these model is proven to be capable to do the job. However, there is a potential flaw in the dataset. The amount of negative outcomes in the training set was significantly smaller than the positive outcome, hence the reason why the F1_score is significanly low. 

A potential recommendation is to decrease the positive outcomes and makes the dataset balance in terms of outcomes.
