```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('/kaggle/input/drug-classification/drug200.csv')
data.head()
```

```python
data['Sex'] = data['Sex'].replace('F', 0)
data['Sex'] = data['Sex'].replace('M', 1)
data['BP'] = data['BP'].replace('LOW', 1)
data['BP'] = data['BP'].replace('NORMAL', 2)
data['BP'] = data['BP'].replace('HIGH', 3)
data['Cholesterol'] = data['Cholesterol'].replace('NORMAL', 0)
data['Cholesterol'] = data['Cholesterol'].replace('HIGH', 1)
data.head()
```

```python
X = data[['Age', 'Sex', 'BP', 'Cholesterol', 'Na_to_K']]
y = data['Drug']

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```
