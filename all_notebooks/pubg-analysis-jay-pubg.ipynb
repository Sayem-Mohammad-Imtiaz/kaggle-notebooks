```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as ws
ws.filterwarnings("ignore")
```

```python
print ("LET's GET BACK TO THE SAFEZONE")
```

```python
df = pd.read_csv("/kaggle/input/pubg-weapon-stats/pubg-weapon-stats.csv")
```

```python
df.head()
```

```python
def clean_name(name):
    return name.strip().lower().replace(" ", "_")
df.rename(columns=clean_name, inplace = True)
```

```python
# Different Weapon Type
sns.set()
plt.figure(figsize=(20,5))
sns.countplot (df["weapon_type"], palette="rainbow")
plt.show()
# Most occuring assualt Rifle
```

```python
# Diffrent types of ammuniation
sns.set()
plt.figure(figsize=(6,6))
sns.countplot (df["bullet_type"])
plt.show()
# Most occuring assualt Rifle
```

# Most important is the size of Magazine Capacity

### While Playing the TPS (Third Person Shooter) game like PUBG or CSGO you always need the  gun with good Magazine size. It will help you to shoot the player on the long range with continous  bullets.

### During the Close Combats with enemy or  1 Vs 1 battle  the large Magazine helps us to kill the enemy with spray down of the guns

```python
sns.set()
plt.figure(figsize = (10,10))
sns.countplot (df.magazine_capacity, palette="inferno")
plt.show()

# Most commonly a gun has 30 bullets 
```

# Another important aspect is the " Damage of the Gun "

It will decide the amount of damage to the player once the bullet of certain weapon is fired

```python
temp = df.groupby("weapon_type")["damage"].count().reset_index()
```

```python
# The weapon damage according to the size of rifle
plt.figure(figsize = ( 6,6))
sns.barplot(y="weapon_type", x ="damage",data = temp)
plt.show()
```

# Here we can conclude that the Assault Rifle has the  higher damage 
## like 1. AKM, M416, SCAR-L and  many more

#### But in Rifles which has the highest damage??? let's find 


```python
most_damaging_weapon = df[df["weapon_type"] == "Assault Rifle"][["damage", "weapon_name"]].sort_values("damage", ascending  = False)
most_damaging_weapon
```

```python
plt.figure(figsize = (6,6))
sns.barplot(y="weapon_name", x ="damage",data = most_damaging_weapon)
plt.show()
```

**Most weapons are optimal for the close combats  or the short range fights **

### Now we will inspect the averagae bullet speed

```python
speed = df.groupby("weapon_type")["bullet_speed"].mean().reset_index().dropna()
speed
```

```python
fig, ax = plt.subplots(figsize=(16,10), dpi= 80)
ax.vlines(x=speed["weapon_type"], ymin=0, ymax=speed["bullet_speed"], color='firebrick', alpha=0.7, linewidth=2)
ax.scatter(x=speed["weapon_type"], y=speed["bullet_speed"], s=75, color='firebrick', alpha=0.7)
#Annotate Title
ax.set_title('Average Speed of the Bullet per weapon type', fontdict={'size':22})
ax.set_ylabel('Average Speed ')
ax.set_xticks(speed.weapon_type)
plt.show()
```

# **Light Weight Machine Guns has high velocity of the bullets per unit area**

# Now we will analyze the average fire rate of the each weapon type

```python
fire_rate = df.groupby("weapon_type")["rate_of_fire"].mean().reset_index().dropna()

fire_rate
```

```python
# Average Fire Rate
plt.figure(figsize=(20,5))
plt.title("Average Firing rate per weapon type")
sns.barplot (x = "weapon_type", y = "rate_of_fire", data = fire_rate, palette="rainbow")
plt.show()
```

####  We can clearly see that the  Assault rifle has the lowest fire rate

```python
plt.rcParams['figure.figsize'] = (8, 6)
plt.rcParams['figure.dpi'] = 150
sns.scatterplot(x=df.bullet_speed, y = df.damage)
plt.show()
```

### By primary intuiation and also by analysis we shown that the increase in bullet speed does the higher damage
