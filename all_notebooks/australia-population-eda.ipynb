```python
import os
import numpy as np
import pandas as pd
import random
import seaborn as sns
import datetime as datetime
import matplotlib.dates as dates
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
data0 = pd.read_csv("../input/australia-population-births-and-deaths/Australia Population.csv",sep=';')
data0[0:5]
```

```python
print(data0.shape)
```

```python
print(data0['Date'][0:6])
print()
print(data0['Date'][-6:])
```

```python
data0['Year']=data0['Date'].apply(lambda x:int(x[-4:]))
data0['Year']
```

```python
data1=data0.groupby('Year').sum()
data2=data1[1:-1]
data2[-5:]
```

```python
data2.columns
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=data2.index,y=data2['Deaths'],name='Deaths'),secondary_y=False,)
fig.add_trace(go.Scatter(x=data2.index,y=data2['Births'],name='Births'),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="Births and Deaths")
fig.update_xaxes(title_text="Year")
fig.update_yaxes(title_text="Number",secondary_y=False)
fig.show()
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=data2.index,y=data2['Overseas Departures'],name='Overseas Departures'),secondary_y=False,)
fig.add_trace(go.Scatter(x=data2.index,y=data2['Overseas Arrivals'],name='Overseas Arrivals'),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="Overseas Departures and Arrivals")
fig.update_xaxes(title_text="Year")
fig.update_yaxes(title_text="Number",secondary_y=False)
fig.show()
```

```python
fig=make_subplots(specs=[[{"secondary_y":False}]])
fig.add_trace(go.Scatter(x=data2.index,y=data2['Interstate Departures'],name='Interstate Departures and Arrivals'),secondary_y=False,)
fig.update_layout(autosize=False,width=700,height=500,title_text="Interstate Departures and Arrivals")
fig.update_xaxes(title_text="Year")
fig.update_yaxes(title_text="Number",secondary_y=False)
fig.show()
```
