```python
plt.figure(figsize= (25,7))
plt.style.use("fivethirtyeight")
sns.lineplot(x='Area',y='Value',data=data)
plt.xticks(rotation=90)
plt.show()
```

# China ,China, mainland ,this high produce


```python
data.groupby(data['Area']).sum().sort_values(by='Value',ascending=False)
```

```python
top=data[(data['Area']=='China' ) | (data['Area']=='China, mainland' )]
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=top,x="Year",y="Value",marker='o',hue="Area")
plt.show()
```

```python
plt.figure(figsize=(20,7))
plt.style.use('ggplot')
sns.lineplot(data=data,x="Year",y="Value",marker='o')
```

```python
data.groupby('Year').sum().sort_values(by='Value',ascending=False).head()
```

# Top 3 Years

```python
top_year=data[(data['Year']==2019 ) | (data['Year']==2018 )|(data['Year']==2017 )]
```

```python
plt.figure(figsize=(25,7))
sns.lineplot(data=top_year,x="Area",y="Value",marker='*',hue="Year")
plt.xticks(rotation=90)
plt.show()
```
