```python
import pandas as pd
import numpy as np 
df = pd.read_csv("/kaggle/input/drug-classification/drug200.csv")
df.head()
```

```python
## We fixed the skewness with log 
sns.distplot(np.log(df["Na_to_K"]))
df["Na_to_K"] = np.log(df["Na_to_K"])
```

```python
df["BP"].value_counts()
```

```python
df["Cholesterol"].value_counts()
```

```python
## I don't want to label encode this features because these are ordinal values
BP_map= {
    "HIGH":3,
    "NORMAL":2,
    "LOW":1
}
Cholesterol_map={ 
    "HIGH":1,
    "NORMAL":0
    
}
```

```python
df.loc[:,'BP'] = df['BP'].map(BP_map)
df.loc[:,'Cholesterol'] = df['Cholesterol'].map(Cholesterol_map)
df.head()
```

```python
## I get dummies to the Sex feature because it is not ordinal and the effect of each individual should have same amount
df = pd.concat([df["Drug"],(pd.concat([pd.get_dummies(df["Sex"]),df.drop("Drug",axis=1)],axis=1).drop("Sex",axis=1))],axis=1)
df.head()
```
