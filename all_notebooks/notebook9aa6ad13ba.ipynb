```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
do = pd.read_csv('../input/housesalesprediction/kc_house_data.csv')
do['log_sqft_lot'] = np.log(1+do['sqft_lot'])
do['log_price'] = np.log(1+do['price'])
```

```python
do.head(2)
```

```python
do1 = do[ ['log_sqft_lot', 'floors', 'log_price'] ]
do2 = do['zipcode']
```

```python
dummies = pd.get_dummies(do2)
```

```python
mrg = pd.concat([do1,dummies],axis='columns')
```

```python
final = mrg.drop([98001], axis='columns')
```

```python
N = final.drop('log_price', axis='columns')
M = final['log_price']
```

```python
trn_do = N[0:17000]
test_do  = N[17000:]
```

```python
trn_N = trn_do
trn_M = M[0:17000].values
```

```python
test_N = test_do.values
test_M = M[17000:].values
```

```python
print( trn_N.shape, test_N.shape )
print( trn_M.shape, test_M.shape )
```

```python
import keras
from keras import layers
from keras.models import Model
from keras.optimizers import SGD
```

```python
input1 = layers.Input( shape=(trn_N.shape[1],) , name='entree') 
n    = layers.Dense(200, name='cachee') ( input1 )
active = layers.Activation('relu') (n)
output1 = layers.Dense(1, name='sortie') (active) 
```

```python
model  = Model( input1, output1 )
```

```python
model.summary()
```

```python
opt1 = SGD() 
```

```python
model.compile( optimizer=opt1, loss='mse')
```

```python
model.fit( trn_N, trn_M, epochs=5 )
```

```python
predictions = model.predict(test_N)
```
