# Importing Data

# Data Analysis and Visualization

### Comparing County with Area in Square Km

```python
plt.figure(figsize=(8, 20))

plt.barh(population["County"],population["Area in Square Km."],color="blue")
plt.xlabel("Area in Square Km")
plt.ylabel("County")
plt.show()
```

> Turkana and Marsabit are the largest counties in terms of area in square km 

### Comparing Total Population per County

```python
plt.figure(figsize=(8, 20))

plt.barh(population["County"],population["Total Population"],color="blue")
plt.xlabel("Total Population")
plt.ylabel("County")
plt.show()
```

> Nairobi has the clear lead as the most populated county

### Comparing Population of Females and Males in Kenya

```python
y = np.array([19417639, 19192458])
mylabels = ["Females", "Males"]

plt.pie(y, labels = mylabels, autopct='%1.1f%%')
plt.title(" Comparison of Female and Male Population in Kenya")
plt.show()
```

> Almost equal number of females and males in Kenya

### Comparing Number of Households and Total Population

```python
plt.figure(figsize=(15,15))

plt.scatter(population["Total Population"], population["No of HouseHolds"], label= population["County"], color="blue")
plt.xlabel("Total Population per County")
plt.ylabel("No of HouseHolds")
plt.show()
```

> The scatter plot shows that as the population increases so does the number of households
