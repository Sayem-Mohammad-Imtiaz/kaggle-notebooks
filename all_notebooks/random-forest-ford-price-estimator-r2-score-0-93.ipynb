```python
import numpy as np 
import pandas as pd 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR # very bad!
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error
sns.set(font_scale=1.3) 
```

```python
df_ford = pd.read_csv('/kaggle/input/used-car-dataset-ford-and-mercedes/ford.csv')
df_ford.head(3)
```

```python
sns.pairplot(df_ford);
```

```python
X = df_ford.drop(['price'],axis=1)
y = df_ford['price']
```

```python
set(X['transmission']),set(X['fuelType'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
df_test = X_test.copy()
X_train.shape,X_test.shape
```

```python
cat_cols = ['model','transmission','fuelType']#'model',
oh_enc = OneHotEncoder(sparse=False,handle_unknown='ignore')
X_train_ = oh_enc.fit_transform(X_train[cat_cols])
X_test_ = oh_enc.transform(X_test[cat_cols])
X_train = X_train.drop(cat_cols,axis=1)
X_test = X_test.drop(cat_cols,axis=1)
X_train = np.concatenate([X_train,X_train_],axis=1)
X_test = np.concatenate([X_test,X_test_],axis=1)
X_train.shape,X_test.shape
```

```python
clf = LinearRegression().fit(X_train,y_train)
y_pred = clf.predict(X_test)
print('R2 score:',r2_score(y_test,y_pred))
print('MAE:',mean_absolute_error(y_test,y_pred))
```

```python
clf = RandomForestRegressor(n_estimators=1000,max_depth=29,min_samples_split=9,max_samples=0.6).fit(X_train,y_train)
y_pred = clf.predict(X_test)
print('R2 score:',r2_score(y_test,y_pred))
print('MAE:',mean_absolute_error(y_test,y_pred))
```

```python
cols = df_test.columns
df_test['diff'] = pd.Series((y_test-y_pred))
df_test['price'] = y_test
sns.scatterplot(data=df_test,x='price',y='diff');
```
