```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("../input/restaurant-business-rankings-2020/Future50.csv")
df.head(10)
```

```python
df.info()
```

```python
df['Franchising'] = df['Franchising'].map({'Yes': 1, 'No': 0})

df.info()
df['Franchising']
```

```python
#check for missing values
df.isnull().sum()
df
```

```python

x=df.drop('Franchising',axis=1)
x=x[['Sales','Units','Unit_Volume']]
y=df['Franchising']
print(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,stratify=y,random_state=42)
x_train,y_train,x_test,y_test
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)

# Train the model using the training sets
model.fit(x_train,y_train)

#Predict Output
predicted= model.predict(x_test) # 0:Overcast, 2:Mild
print(predicted)
```

```python
from sklearn import svm
clf = svm.SVC()
clf.fit(x_train, y_train)
yhat=clf.predict(x_test)
yhat
```

```python
from sklearn.metrics import accuracy_score
knn_accuracy=accuracy_score(y_test,predicted)
svm_accuracy=accuracy_score(y_test,yhat)
print("KNN accuracy: ",'%.3f'%(knn_accuracy*100),"%")
print("SVM accuracy: ",svm_accuracy*100,"%")
```

```python
#ROC Curve for KNN
import sklearn.metrics as metrics
# calculate the fpr and tpr for all thresholds of the classification
probs = model.predict_proba(x_test)
preds = probs[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, predicted)
roc_auc = metrics.auc(fpr, tpr)
from ggplot import *
df = pd.DataFrame(dict(fpr = fpr, tpr = tpr))
ggplot(df, aes(x = 'fpr', y = 'tpr')) + geom_line() + geom_abline(linetype = 'dashed')
```

```python
#ROC Curve for SVM
import sklearn.metrics as metrics
# calculate the fpr and tpr for all thresholds of the classification
probs = model.predict_proba(x_test)
preds = probs[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, yhat)
roc_auc = metrics.auc(fpr, tpr)
from ggplot import *
df = pd.DataFrame(dict(fpr = fpr, tpr = tpr))
ggplot(df, aes(x = 'fpr', y = 'tpr')) + geom_line() + geom_abline(linetype = 'dashed')
```

```python
from sklearn.metrics import roc_auc_score

# auc scores
auc_score_knn = roc_auc_score(y_test, predicted)
auc_score_svm = roc_auc_score(y_test,yhat)
print("AUC Score of KNN =",'%.3f'%auc_score_knn)
print("AUC Score of SVM =",'%.3f'%auc_score_svm)
```
