**Loading data**

```python
data = pd.read_csv("/kaggle/input/pakistans-top-250-youtube-channels/Pakistan Top 250 YouTube Channels.csv")

# displaying top 10 records
data.head(10)
```

```python
data.columns
```

**Space after column name `Twitter` will cause problem So fixing this**

```python
data.rename({'Twitter '  : "Twitter"}, axis = 1, inplace = True)
```

```python
# displaying columns name
data.columns
```

**Seperating `Channel's name`, `Category` and `Subscriber`**

```python
data[["ChannelName", "Category", "Subscriber"]] = data['Twitter'].str.split(',', expand = True)
```

```python
# displaying top 10 records
data.head()
```

**Dropping `Top #` and `Twitter` column from `data`**

```python
data.drop(["Top #", "Twitter"], axis = 1, inplace = True)
```

```python
display(data)
```

```python
data.dtypes
```

```python
def unit_to_float(x):
    if type(x) == float or type(x) == int:
        return x
    if 'K' in x or 'k' in x: 
        if len(x) > 1:
            return float(x.replace('K', '')) * 1000
        return 1000.0
    if 'M' in x or 'm' in x:
        if len(x) > 1:
            return float(x.replace('M', '')) * 1000000
        return 1000000.0
    if 'B' in x or 'b' in x:
        return float(x.replace('B', '')) * 1000000000
    return 0.0
```

```python
data['Subscriber'] = data['Subscriber'].apply(unit_to_float)
```

```python
data.head()
```

#### **Now checking what most Pakistani are watching**

```python
plt.figure(figsize = (25,10))
sns.barplot(x = "Category", y = "Subscriber",
            data = data,
            ci = None,
           edgecolor = "mediumblue",
           lw = 4)

plt.title("MOST VIEWED CATEGORY IN PAKISTAN\n", size = 30)
plt.xticks(rotation = 45)
plt.xlabel("Category\n", size = 20)
plt.ylabel("Subscribers\n", size = 20)

plt.show()
```

#### **Now check for the top 10 `Entairteinment` YT channels**

```python
entairteinment_data = data[data['Category'] == "Entertainment"]

entairteinment_data.head(10)
```

```python
# sort subscriber's column in descending order
entairteinment_data.sort_values("Subscriber", ascending=False, inplace=True)
```

```python
# displaying top 10 Entertainment YT channel in Pakistan
entairteinment_data.head(10)
```

```python
plt.figure(figsize = (15,8))
sns.barplot(x = "Subscriber", y = "ChannelName",
            data = entairteinment_data.head(10), palette="rainbow")
plt.title("TOP 10 PAKISTANI ENTAIRTEINMENT YOUTUBE CHANNELS\n", size = 15)
plt.show()
```
