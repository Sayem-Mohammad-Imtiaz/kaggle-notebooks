```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
from sklearn import preprocessing
encoder = preprocessing.LabelEncoder()
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR

df = pd.read_csv('../input/flight-take-off-data-jfk-airport/M1_final.csv')
df.head()
```

```python
ans = df.iloc[:,:-1]
```

# DATA FITTING

```python
ans['OP_UNIQUE_CARRIER'] = encoder.fit_transform(ans['OP_UNIQUE_CARRIER'].astype(str))
ans['TAIL_NUM'] = encoder.fit_transform(ans['TAIL_NUM'].astype(str))
ans['DEST'] = encoder.fit_transform(ans['DEST'].astype(str))
ans['Wind'] = encoder.fit_transform(ans['Wind'].astype(str))
ans['Condition'] = encoder.fit_transform(ans['Condition'].astype(str))
ans.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(ans,df['TAXI_OUT'],test_size=0.1)
```

# LINEAR REGRESSOR(LE)

```python
linreg = LinearRegression()
linreg.fit(x_train,y_train)
y_predicted1=linreg.predict(x_test)

r1=mean_squared_error(y_test, y_predicted1)**0.5
r1
```

# RIDGE REGRESSION(LE)

```python
RR =Ridge(alpha = .05, normalize =True)
RR.fit(x_train,y_train)
ypred2=RR.predict(x_test)
r2=mean_squared_error(y_test, ypred2)**0.5
r2
```

# LASSO REGRESSOR(LE)

```python
Lasreg=Lasso(alpha = .05 , normalize =True)
Lasreg.fit(x_train,y_train)
ypred3= Lasreg.predict(x_test)


r3=mean_squared_error(y_test, ypred3)**0.5
r3
```

# KNN(LE)

```python
KNN=KNeighborsRegressor(n_neighbors = 101)
KNN.fit(x_train,y_train)
ypred4= KNN.predict(x_test)

r4=mean_squared_error(y_test,ypred4)**0.5
r4
```

# SVM(LE)

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
svm=SVR()
a = x_train.copy()
b = x_test.copy()
stda = StandardScaler()
stdb = StandardScaler()
a = stda.fit_transform(a)
b = stdb.fit_transform(b)
svm.fit(a,y_train)
ypred5 = svm.predict(b)
r5 = mean_squared_error(y_test,ypred5)**0.5
r5
```

# NAIVE BAYES(LE)

```python
from sklearn.linear_model import BayesianRidge

bay = BayesianRidge()
bay.fit(x_train, y_train)
ypred6 = bay.predict(x_test)
r6 = mean_squared_error( y_test,ypred6)**0.5
r6
```

# RANDOM FORREST(LE)

```python
from sklearn.ensemble import RandomForestRegressor


forrest=RandomForestRegressor()
forrest.fit(x_train,y_train)
ypred7= forrest.predict(x_test)


r7=mean_squared_error(y_test,ypred7)**0.5
r7
```

# LGBM(LE)

```python
from lightgbm import LGBMRegressor
LGBM= LGBMRegressor()

LGBM=BayesianRidge()
LGBM.fit(x_train,y_train)
ypred8= LGBM.predict(x_test)


r8=mean_squared_error(y_test, ypred8)**0.5
r8
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(handle_unknown='ignore')

# col1=pd.DataFrame(enc.fit_transform(features[['OP_UNIQUE_CARRIER']]).toarray())
# features=features.join(col1)
# features
# features=pd.merge(features,col1, left_index=True, right_index=True)
# features
ans=pd.get_dummies(ans,columns=["OP_UNIQUE_CARRIER","DEST","Wind","Condition"])
```

```python
ans.drop(["TAIL_NUM"],inplace =True,axis=1)
from sklearn.model_selection import train_test_split

```

```python
X_train,X_test,Y_train,Y_test=train_test_split(ans,df['TAXI_OUT'],test_size=0.1)
```

# LINEAR REGRESSOR(OHE)

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

LR=LinearRegression()
LR.fit(X_train,Y_train)
Y_pred=LR.predict(X_test)

O1=mean_squared_error(Y_test, Y_pred)**0.5
O1
```

# RIDGE REGRESSOR(OHE)

```python
from sklearn.linear_model import Ridge

RR =Ridge(alpha = .05, normalize =True)
RR.fit(X_train,Y_train)
Y_pred2=RR.predict(X_test)


O2=mean_squared_error(Y_test, Y_pred2)**0.5
O2
```

# LASSO REGRESSOR(OHE)

```python
from sklearn.linear_model import Lasso

Lasreg=Lasso(alpha = .05 , normalize =True)
Lasreg.fit(X_train,Y_train)
Y_pred3= Lasreg.predict(X_test)


O3=mean_squared_error(Y_test, Y_pred3)**0.5
O3
```

# KNN(OHE)

```python
from sklearn.neighbors import KNeighborsRegressor

KNN=KNeighborsRegressor(n_neighbors = 101)
KNN.fit(X_train,Y_train)
Y_pred4= KNN.predict(X_test)

O4=mean_squared_error(Y_test, Y_pred4)**0.5
O4
```

# SVM(OHE)

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
c = X_train.copy()
d = X_test.copy()
s1 = StandardScaler()
s2 = StandardScaler()
c = s1.fit_transform(c)
d = s2.fit_transform(d)
svm  = SVR()
svm.fit(c,Y_train)
Y_pred5 = svm.predict(d)
O5 = mean_squared_error(Y_pred5, Y_test, squared= False)
O5
```

# NAIVE BAYES(OHE)

```python
from sklearn.linear_model import BayesianRidge
bay = BayesianRidge()
bay.fit(X_train, Y_train)
Y_pred6 = bay.predict(X_test)
O6 = mean_squared_error(Y_pred6, Y_test,squared=False)
O6
```

# RANDOM FORREST(OHE)

```python
from sklearn.ensemble import RandomForestRegressor
forrest = RandomForestRegressor()
forrest.fit(X_train, Y_train)
Y_pred7 = forrest.predict(X_test)
O7 = mean_squared_error(Y_pred7, Y_test, squared = False)
O7
```

# LBGM(OHE)

```python

from lightgbm import LGBMRegressor
light = LGBMRegressor()

light=BayesianRidge()
light.fit(X_train,Y_train)
Y_pred8 = light.predict(X_test)
O8 = mean_squared_error(Y_pred8, Y_test,squared=False)
O8
```
