# KMeans Clustering

```python
column_1 = data.iloc[:,1]
column_2 = data.iloc[:,3]
dictionary = {"ROA":column_1,"ROB":column_2}
data2 = pd.DataFrame(dictionary)
data2.head()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 4)
kmeans.fit(data2)
labels = kmeans.predict(data2)

plt.scatter(data.iloc[:,1],data.iloc[:,3],c = labels)
plt.xlabel("ROA(C) before interest and depreciation before interest")
plt.ylabel("ROA(B) before interest and depreciation after tax")
plt.title("ROA(C) - ROA(B)")
plt.show()
```

### İf we divide our data 4 cluster it will be like that.But why 4 cluster,let's try more of 4 cluster,What we will see.

```python
wcss = []

for each in range(1,10):
    kmeans = KMeans(n_clusters = each)
    kmeans.fit(data2)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,10),wcss,'-o')
plt.xlabel("Count of cluster")
plt.ylabel("WCSS")
plt.show()
```

WCSS : Within cluster sum of squares 

WCSS mean,we will take cluster centers mean and we're gonna calculate distance each of sample,İf you wanna learn more.You can look sklearn documentation


İf we increase cluster count it will be much less but we can't do it.Because if we do our data is going to be overfitting.We don't want it.
So we must take **elbow** count.When we take it,model will more clean and good.That's the literature.

Elbow : When you see an elbow(İt is 2 count of cluster you can see the elbow)

```python
#Again we will create our model with 3 count of cluster
final_kmeans = KMeans(n_clusters = 2)
final_kmeans.fit(data2)
labels = final_kmeans.predict(data2)

plt.scatter(data.iloc[:,1],data.iloc[:,3],c = labels)
plt.xlabel("ROA(C) before interest and depreciation before interest")
plt.ylabel("ROA(B) before interest and depreciation after tax")
plt.title("ROA(C) - ROA(B)")
plt.show()
```

# STANDARDIZATION

* Standardizaton is important for both supervised and unsupervised learning
* Do not forget standardization as pre-processing
* As we already have visualized data so you got the idea. Now we can use all features for clustering.
* We can use pipeline like supervised learning.

```python
data = pd.read_csv("../input/company-bankruptcy-prediction/data.csv")
sub_data = data.drop('Bankrupt?',axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

scalar = StandardScaler()
kmeans = KMeans(n_clusters = 2)

pipe = make_pipeline(scalar,kmeans)
pipe.fit(sub_data)
labels = pipe.predict(sub_data)
```

### And Evaluate

```python
df = pd.DataFrame({'labels':labels,"Bankrupt?":data['Bankrupt?']})
ct = pd.crosstab(df['labels'],df['Bankrupt?'])

print(ct)
```

# HİERARCHY

* vertical lines are clusters
* height on dendogram: distance between merging cluster
* method= 'single' : closest points of clusters

```python
from scipy.cluster.hierarchy import linkage,dendrogram

merg = linkage(sub_data.iloc[200:220,:],method = "single")#because of interrupt we did 200 between 300
dendrogram(merg,leaf_rotation = 90,leaf_font_size = 6)
plt.show()
```
