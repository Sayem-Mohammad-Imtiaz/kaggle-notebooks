***Exploratory Analysis on the Student Performance Dataset***

We will look into some of the factors influencing the passing statuses of students in Math, Reading and Writing Examinations.


*Let's load our libraries...*

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

*Now we read the dataset*

```python
data = pd.read_csv('../input/StudentsPerformance.csv')
    
```

```python
data.describe()
```

```python
data.head(5)
```

*Checking for any null value*

```python
data.isnull().sum()
```

*Lets introduce minimum passing marks to  pass the subject*

```python
passscore = 40
```

```python
data['math_passed']=np.where(data['math score']>=passscore, 'p','f')
data.math_passed.value_counts()
```

*We will drop the columns not necessary for us*

```python
data.drop(['lunch','test preparation course'],axis=1)
```

```python
sns.countplot(x='gender', data=data)
plt.show()
```

```python
sns.countplot(x='math_passed', hue='gender', data= data)
plt.show()
```

```python
data['reading_passed']=np.where(data['reading score']>passscore, 'p','f')
data['writing_passed']=np.where(data['writing score']>passscore, 'p','f')
data.head(5)
```

```python
fig, ax = plt.subplots(2,1,figsize=(12,6))
sns.countplot(x='reading_passed', hue='gender', data= data, ax=ax[0])
sns.countplot(x='writing_passed', hue='gender',data=data, ax=ax[1])
plt.show()
```

*Now let's see how the parental level of education influence the performance of students in different exams*

```python
sns.countplot(x='parental level of education', hue='math_passed', data=data)
plt.show()
```

```python
sns.countplot(x='parental level of education', hue='math_passed', data=data)
plt.show()
```

```python
sns.countplot(x='parental level of education', hue='reading_passed', data=data)
plt.show()
```

```python
sns.countplot(x='parental level of education', hue='writing_passed', data=data)
plt.show()
```
