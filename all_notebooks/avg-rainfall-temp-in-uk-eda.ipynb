```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings(action="ignore")
from dataprep.clean import clean_headers
sns.set_style("darkgrid")
plt.style.use("fivethirtyeight")
```

```python
df = pd.read_csv("/kaggle/input/average-rainfall-and-temperature-in-uk20102019/avg_rainfalltemp in UK - Sheet1.csv")
df.head(10)
```

#### Now, we clean column names so that it would be convenient.

```python
df = clean_headers(df,case="camel")
```

```python
# cleaned columns
df.columns
```

```python
# Percentage of null values
(df.isna().sum()/len(df))*100
```

## EDA

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True,fmt = ".1f",cmap="RdBu",linecolor="black")
plt.title("Correlation Matrix")
```

```python
plt.figure(figsize=(18,8))
sns.pairplot(df,hue="period")
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df["avgRainfallInMm"])
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df["avgTempInCentigrade"],bins=15)
```

```python
plt.figure(figsize=(14,8))
sns.relplot(x="avgTempInCentigrade",y="period",data=df,hue="typeOfPeriod")
```

> **July has highest average temperature.**

```python
plt.figure(figsize=(16,10))
sns.relplot(x="avgRainfallInMm",y="period",data=df,hue="typeOfPeriod")
```

> **Spring has less average rainfall than other Seasons.**

```python
plt.figure(figsize=(10,8))
sns.boxplot(y="avgRainfallInMm",data=df)
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(y = "avgTempInCentigrade",data=df)
```
