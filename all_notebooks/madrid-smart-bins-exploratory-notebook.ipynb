```python
file_path = '/kaggle/input/madrid-smart-bins/papeleras_madrid.csv'
df = pd.read_csv(file_path, delimiter=";")
df.head()
```

```python
df.describe()
```

```python
#Check if there are missing values
df.isnull().sum()
```

```python
s = (df.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables:")
print(object_cols)
```

```python
count = df['alerta'].value_counts()
print(count)
```

```python
llena = (df["alerta"] == "LLENA").sum()
error = (df["alerta"] == "ERROR").sum()
nan = df['alerta'].isnull().sum()
print(llena, error, nan)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
state = ['Llena', 'Error', 'Nan']
freq = [llena, error, nan]
ax.bar(state,freq)
ax.bar_label(ax.containers[0])
plt.show()
```

```python
#No NaNs
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
state = ['Llena', 'Error']
freq = [llena, error]
ax.bar(state,freq)
ax.bar_label(ax.containers[0])
plt.show()
```
