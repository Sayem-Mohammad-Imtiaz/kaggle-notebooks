```python


import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

df=pd.read_csv('../input/world-happiness-report/2019.csv')





```

```python
df.columns=df.columns.str.replace('Country or region','Country')
df.columns=df.columns.str.replace('GDP per capita','Economy')
df.columns=df.columns.str.replace('Social support','Social_support')
df.columns=df.columns.str.replace('Healthy life expectancy','Health')
df.columns=df.columns.str.replace('Freedom to make life choices','Freedom')
df.columns=df.columns.str.replace('Perceptions of corruption','Perceptions_of_corruption')
```

```python
ulkeler = list(df["Country"].unique())
```

```python
df_ulkeler = pd.DataFrame(ulkeler)
df_ulkeler
```

```python
ulke_mutluluk_orani = []


for i in ulkeler:
    deger = df[df["Country"] == i]
    ulke_mutluluk_skoru = sum(deger.Score)/len(deger)
    ulke_mutluluk_orani.append(ulke_mutluluk_skoru)
```

```python
df = pd.DataFrame({"Country": ulkeler , "Happiness_Ratio" : ulke_mutluluk_orani})
yeni_index = (df["Happiness_Ratio"].sort_values(ascending=False)).index.values
sirali_veri = df.reindex(yeni_index)
```

```python
plt.figure(figsize = (25,15))
sns.barplot(x = sirali_veri["Country"][:50] , y=sirali_veri["Happiness_Ratio"][:50])
plt.xticks(rotation = 90)
plt.xlabel("Country" , fontsize=20 , color = "purple")
plt.ylabel("Average Happiness Score" , fontsize= 20 , color = "purple")
plt.show()
```

```python
ulkeler = list(df["Country"].head(10))
```

```python
ekonomi = []
aile = []
saglik = []
ozgurluk = []


for i in ulkeler:
    deger = df[df["Country"] == i]
    ekonomi.append(sum(deger.Economy)/len(deger))
    aile.append(sum(deger.Social_support)/len(deger))
    saglik.append(sum(deger.Health)/len(deger))
    ozgurluk.append(sum(deger.Freedom)/len(deger))
```

```python
f,ax = plt.subplots(figsize=(20,15))
sns.barplot(x = ekonomi , y=ulkeler , color = "red" , alpha=0.5 , label="Economy")
sns.barplot(x = aile , y=ulkeler , color = "orange" , alpha=0.5,label="SocialSupport")
sns.barplot(x = saglik , y=ulkeler , color = "blue" , alpha=0.5,label="Health")
sns.barplot(x = ozgurluk , y=ulkeler , color = "purple" , alpha=0.5,label="Freedom")


ax.legend(loc="lower right")

plt.ylabel("Country" , fontsize=20, color='violet') 
plt.xlabel("Impact Rate" , fontsize =20, color='violet')
plt.show();
```
