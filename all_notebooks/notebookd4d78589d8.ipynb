```python
nltk.download('stopwords')
```

```python
def process_text(text):
    #remove punctuation
    nopunc = [char for char in text if char not in string.punctuation]
    nopunc = ''.join(nopunc)
    
    #remove stop words
    clean_words = [word for word in nopunc.split() if word.lower() not in stopwords.words('english')]
    
    #Return a list of clean words
    return clean_words
```

```python
data['email'].head().apply(process_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

message_bow = CountVectorizer(analyzer = process_text).fit_transform(data['email'])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(message_bow, data['label'],test_size = 0.20, random_state = 0)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
classifier= MultinomialNB()
```

```python
classifier.fit(X_train,y_train)
```

```python
classifier.score(X_test,y_test)
```
