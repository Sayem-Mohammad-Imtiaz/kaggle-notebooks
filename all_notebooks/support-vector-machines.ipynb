```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("../input/prostate-cancer/Prostate_Cancer.csv")
```

```python
print(df.head(5))
```

```python
df.drop("id", axis = 1, inplace = True)
```

```python
print(df.head(5))
```

```python
classification = {'B' : 0, 'M' : 1}
df['diagnosis_result'] = df['diagnosis_result'].map(classification)
```

```python
print(df.head(5))
```

```python
df.shape
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:, 1:10], df['diagnosis_result'], test_size = 0.3)
```

```python
svc.fit(x_train,y_train)
```

```python
pred = svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```

```python
#change the kernel
parameter = {'C' : [0.1,1,10,100,1000,10000], 'kernel':['rbf'], 'gamma': [1,0.1,0.01,0.001,0.0001,0.00001]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(), parameter, refit=True, verbose = 3)
```

```python
grid.fit(x_train, y_train)
```

```python
grid.best_params_
```

```python
pred = grid.predict(x_test)
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(classification_report(y_test, pred))
```
