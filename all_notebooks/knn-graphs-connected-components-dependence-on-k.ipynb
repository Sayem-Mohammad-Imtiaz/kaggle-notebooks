# What is about 


Study of number of connected components of random KNN graph depending on K and data dimension.

Expectations: 

(1) for K big enough one has only one connected component.
"Big enough" threshold depends on dimension. Simulation shows for D=2  for K = 7 - only one connected component,
for D=3 -  K=6 , D=20 - K = 4 . That is related to percolation threshold. 

(2) For K=1 situation is different and discussed separetely.
The fact is that number of components grows linearly with respect to size of the sample.
The coefficient in front of that linear growth is well-established for D=2 and proposed for higher D:
https://cstheory.stackexchange.com/questions/47034/number-of-connected-components-of-a-random-nearest-neighbor-graph/47037#47037
And notebook:
https://www.kaggle.com/alexandervc/study-connected-components-of-random-nn-graph
Simulation and theoretic proposal agrees in some cases and disagrees in other.

# Original script from the first question on TCS

https://cstheory.stackexchange.com/questions/47034/number-of-connected-components-of-a-random-nearest-neighbor-graph

```python
!pip install python-igraph
!pip install cairocffi
import igraph

import time
from sklearn.neighbors import NearestNeighbors
import numpy as np

t0 = time.time()

dim = 20
n_sample = 10**4

for i in range(10): # repeat simulation 10 times to get more stat
  X = np.random.rand(n_sample, dim)
  nbrs = NearestNeighbors(n_neighbors=2, algorithm='ball_tree', ).fit(X)
  distances, indices = nbrs.kneighbors(X)
  g = igraph.Graph( directed = True )
  g.add_vertices(range(n_sample))
  g.add_edges(indices )
  g2 = g.as_undirected(mode = 'collapse')
  r = g2.clusters()
  print(len(r),len(r)/n_sample*100 , time.time() - t0)
```
