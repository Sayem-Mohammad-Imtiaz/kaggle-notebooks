```python
import pandas as pd 
data=pd.read_csv('../input/fraud-detection-bank-dataset-20k-records-binary/fraud_detection_bank_dataset.csv')
data.head(5)
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor

y=data['targets']
x=data.drop('targets',axis=1)
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
param_grid = {'n_estimators': [1, 30,100], 'max_features': ['auto', 'log2'], 'max_depth': [4,6, 8,10]}
clf_rf = RandomForestClassifier(random_state=5)

cv_model=GridSearchCV(estimator=clf_rf,param_grid=param_grid,cv=10,scoring='recall',n_jobs=-1)
cv_model.fit(xtrain,ytrain)
cv_model.best_params_


```

```python
from sklearn.metrics import accuracy_score 
best_rf=RandomForestClassifier(random_state=5,max_depth=4,max_features='auto',n_estimators=1)
best_rf.fit(xtrain,ytrain)
pred=best_rf.predict(xtest)
print(f'the best accuracy is {round(accuracy_score(ytest,pred)*100,4)}')
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,pred))
```
