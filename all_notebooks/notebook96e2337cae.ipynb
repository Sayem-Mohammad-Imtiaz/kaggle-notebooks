```python
# Let's combine the both data
data=amazon.merge(flipkart,left_on="amazon_isbn-10",right_on='flipkart_isbn10',how='inner')
```

```python
# drop the null values
data=data.dropna()
```

```python
# check the null values are removed or not
data.isnull().sum()
```

```python

data.columns
```

```python
# drop the unwanted columns
data=data.drop(['amazon_isbn-10','flipkart_isbn10','flipkart_title','amazon_author'],axis=1)
```

```python
# Rename the Column name
data=data.rename(columns={'amazon_title':'Book Name','amazon_reviews count':'amazon_ratings count','flipkart_stars':'flipkart_rating',
                         'flipkart_author':'Book Author'})
```

```python
data.drop_duplicates()
```

```python
data=data.drop_duplicates()
```

```python
data=data.reset_index()
```

```python
data.head()
```

```python
# Difference between amazon and Flipkart ratings
plt.figure(figsize=(16,8))
plt.hist(x=data['amazon_rating'],bins=150,color='r',label='Amazon')
plt.hist(x=data['flipkart_rating'],bins=150,color='y',label='Flipkart')
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.title('Amazon V/S Flipkart rating')
plt.legend(loc=1)
plt.grid()
plt.show()
```

```python
# Ratingcounts Difference between Amazon and Flipkart
plt.figure(figsize=(16,8))
plt.hist(data['amazon_ratings count'],edgecolor='r',bins=150,alpha=0.5,fc=(0,0,1,0.7),label='Amazon Rating',range=[0,1000])
plt.hist(data['flipkart_ratings count'],color='k',bins=150,fc=(0,0,1,0.7),label='Flipkart Rating',range=[0,1000])
plt.legend()
plt.xlabel('Ratings')
plt.ylabel('Frequency')
plt.title('Amazon V/S Flipkart Rating Counts')
plt.show()
```

```python
# price defference between Amazon and Flipkart
plt.figure(figsize=(16,8))
plt.hist(x=data['amazon_price'],bins=150,color='r',label='Amazon Price')
plt.hist(x=data['flipkart_price'],bins=150,color='c',alpha=0.5,label='Flipkart Price')
plt.xlabel('Prices')
plt.ylabel('Frequency')
plt.title('Amazon V/S Flipkart Price')
plt.legend(loc=1)
plt.grid()
plt.show()
```

```python
# Amazon rating V/S price plot
sns.jointplot(y='amazon_price',x='amazon_rating',data=data)
plt.show()
```

```python
# flipkart rating V/S price 
sns.jointplot(y='flipkart_price',x='flipkart_rating',data=data)
plt.show()
```
