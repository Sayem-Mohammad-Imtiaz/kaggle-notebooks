Since the movement of the DOGE coin price seems to be "manipulated" somehow, there may be a time-anomaly, which one can take advantage of to make money:D

Here our strategy is to buy & sell on Friday. Yeah that's it.

# Backtest
Let's perform a backtest to verify our strategy.

Here we just buy the coin on Friday (and close-sell at the end of the day).

```python
def backtest(df, day=1):
    """
    Every time buy DOGE with 1000 USD
    """
    
    usd = 1000
    profits = []
    for i in range(len(df)):
        if df['dayofweek'].values[i] == day:
            # buy
            pos_volume = usd / df['Open'].values[i]
            profits.append(pos_volume * (df['Close'].values[i] - df['Open'].values[i]))
    return profits

def plot_profits(profits):
    profits = np.array(profits)
    plt.plot(np.cumsum(profits))
    winp = np.sum(profits > 0) / len(profits)
    plt.xlabel('week')
    plt.ylabel('profits (USD)')
    plt.title(f'P(win) = {winp:.3f}')
    
profits = backtest(df, day=4)
plot_profits(profits)
```
