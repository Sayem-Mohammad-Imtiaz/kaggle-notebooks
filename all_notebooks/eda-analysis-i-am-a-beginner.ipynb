# Import Express from plotly for interactive Graphs

```python
import plotly.express as px
```

# Checking For the null values in the Dataset 

```python
df.isnull().sum()
```

# **Top 5 most infected States in India**

```python
px.bar(df.iloc[0:5],x='State/UTs',y='Total Cases',color='State/UTs',template='plotly_dark',title='Top 5 Infected States')
```

# Top 5 States have most No. of Active Cases

```python
Active_most=df.sort_values(by='Active',ascending=False).iloc[0:5]
```

```python
px.bar(Active_most,x='State/UTs',y='Active',color='State/UTs',template='plotly_dark',title='Top 5 Active Cases States')
```

# Most Deaths

```python
df.sort_values(by='Deaths',ascending=False).iloc[0:5]
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='State/UTs',y='Deaths',data=df.sort_values(by='Deaths',ascending=False).iloc[0:5])
```

# Highest Active to Total Cases Ratio

```python
df.sort_values(by='Active Ratio',ascending=False).iloc[:5]
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='State/UTs',y='Active Ratio',data=df.sort_values(by='Active Ratio',ascending=False).iloc[:5])
```

```python
sns.regplot(x='Total Cases',y='Deaths',data=df)
```
