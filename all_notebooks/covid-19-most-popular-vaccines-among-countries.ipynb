```python
#importing essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
#reading dataset
df= pd.read_csv("../input/covid-world-vaccination-progress/country_vaccinations.csv")
```

```python
df.head()
```

```python
#keep unique record by country
df = df.drop_duplicates('country', keep ='last')

#we leave only the columns of country and vaccines
df= df.drop(['iso_code','date','total_vaccinations','people_vaccinated','people_fully_vaccinated','daily_vaccinations_raw','daily_vaccinations','total_vaccinations_per_hundred','people_vaccinated_per_hundred','people_fully_vaccinated_per_hundred','daily_vaccinations_per_million','source_name','source_website'], axis=1)
print(df)
```

```python
#we separate the different vaccines used by country
df.vaccines.str.split(pat=",", expand=True)
```

```python
#we separate the columns and rename them to be the same
df0=df.vaccines.str.split(pat=",", expand=True).drop([1,2,3,4], axis=1)
df1=df.vaccines.str.split(pat=",", expand=True).drop([0,2,3,4], axis=1)
df2=df.vaccines.str.split(pat=",", expand=True).drop([0,1,3,4], axis=1)
df3=df.vaccines.str.split(pat=",", expand=True).drop([0,1,2,4], axis=1)
df4=df.vaccines.str.split(pat=",", expand=True).drop([0,1,2,3], axis=1)

df0.columns=['vaccines']
df1.columns=['vaccines']
df2.columns=['vaccines']
df3.columns=['vaccines']
df4.columns=['vaccines']

print(df0)
print(df1)
print(df2)
print(df3)
print(df4)
```

```python
#we join the columns
frames = [df0,df1,df2,df3,df4]
result = pd.concat(frames)

#we delete the empty records
result = result[result['vaccines'].notna()]

#we clean the names by removing the spaces
result['vaccines'] = result['vaccines'].str.strip()
print(result)  
```

```python
#we created a graph with a count of countries that use each vaccine
sns.set_style("whitegrid")
plt.figure(figsize= (8,8))
ax= sns.countplot(y='vaccines', data=result, order= result['vaccines'].value_counts().index)
ax.set_xlabel("Most popular vaccines among countries")
plt.show()
```
