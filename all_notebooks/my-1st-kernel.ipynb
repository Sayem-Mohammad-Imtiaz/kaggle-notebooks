We have information on various diseases and their amount from 1916 to 2010. Let's find out what state was **the most cases of rubella** diseases. 

```python
%matplotlib inline
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt

plt.style.use('ggplot') 
plt.rcParams['figure.figsize'] = (10, 5) # some styles for our graph
```

```python
df = pd.read_csv("../input/rubella.csv")
```

```python
df.head()
```

```python
a = df[['state', 'cases']]
a.head()
```

```python
b = a['state'].value_counts()

plt.xlabel(u'State')
plt.ylabel(u'Number of cases')
plt.title(u'Rubella in USA')
b.plot(kind='bar')
```
