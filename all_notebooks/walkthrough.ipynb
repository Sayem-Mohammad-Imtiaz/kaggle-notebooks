```python
#importing the file
import io
import pandas as pd
df2 = pd.read_csv('/kaggle/input/edmunds-car-review/Review.csv')
```

```python
#Peek at the dataset
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)


df2.head()
```

```python
#Number of reviews per company
df2["Company"].value_counts()
```

```python
# Graph for Companies with more than 5000 reviews 
fig, ax = plt.subplots(figsize=(25,10)) 
data = df2['Company'].value_counts()
points = data[data>5000].index 
frequency = data[data>5000].values 

ax.bar(points, frequency) 
ax.set_title('Company Reviews') 
ax.set_xlabel('Reviews') 
ax.set_ylabel('Company')
```

```python
# Graph for Model with more than 2500 reviews 
fig, ax = plt.subplots(figsize=(25,10)) 
data = df2['Model'].value_counts()
points = data[data>2500].index 
frequency = data[data>2500].values 

ax.bar(points, frequency) 
ax.set_title('Model Reviews') 
ax.set_xlabel('Reviews') 
ax.set_ylabel('Company')
```
