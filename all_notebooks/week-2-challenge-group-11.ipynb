# Catergorical variables 
1. make buckets 2. Hotencode / dummy categorical variables

```python
### SpecialWeek
# Make weeknumber
#df['Week'] = pd.to_datetime(df['Date']).dt.week

ChristmasDays = ['2010-12-31','2011-12-30','2012-12-28','2013-12-27']
ThanksgivingDays = ['2010-11-26','2011-11-25','2012-11-23','2013-11-29']
SpecialDays = ['2010-12-31','2011-12-30','2012-12-28','2013-12-27','2010-11-26','2011-11-25','2012-11-23','2013-11-29']

# Make christmas
df['Christmas'] = np.where(df['Date'].isin(ChristmasDays), 1,0)
df['Thanksgiving'] = np.where(df['Date'].isin(ThanksgivingDays), 1,0)
df['NotSpecial'] = np.where(~df['Date'].isin(SpecialDays), 1,0)

#df[['Date','Christmas']] .head(100)
#df[df['Date'] == '2010-12-31']
#df.groupby('Thanksgiving').Thanksgiving.count()
```

```python
# IsHoliday
df['IsHoliday1'] = np.where(df['IsHoliday'] == True, 1,0)
df[['IsHoliday','IsHoliday1']].head()
```

```python
# Type
dummies = pd.get_dummies(df['Type'])

# Add dummies to df
df = pd.concat([df,dummies],axis=1)

#remove original column
del df['Type']
```

```python
# Store
dummies = pd.get_dummies(df['Store'])

# Add dummies to df
df = pd.concat([df,dummies],axis=1)
```

```python
# Department
dummies = pd.get_dummies(df['Dept'])

# Add dummies to df
df = pd.concat([df,dummies],axis=1)
```

```python
### 1. Check if linear or clip make buckets 2. standardize it  
# Unemployment rate
df['Unemployment'] = (df['Unemployment'] - df['Unemployment'].mean())/(df['Unemployment'].std())

# Markdown -> check whether there are catergories = count distinct
# fill missing values
df['MarkDown1'] = df['MarkDown1'].fillna(0)
df['MarkDown2'] = df['MarkDown2'].fillna(0)
df['MarkDown3'] = df['MarkDown3'].fillna(0)
df['MarkDown4'] = df['MarkDown4'].fillna(0)
df['MarkDown5'] = df['MarkDown5'].fillna(0)

# standardize markdown
df['MarkDown1'] = (df['MarkDown1'] - df['MarkDown1'].mean())/(df['MarkDown1'].std())
df['MarkDown2'] = (df['MarkDown2'] - df['MarkDown2'].mean())/(df['MarkDown2'].std())
df['MarkDown3'] = (df['MarkDown3'] - df['MarkDown3'].mean())/(df['MarkDown3'].std())
df['MarkDown4'] = (df['MarkDown4'] - df['MarkDown4'].mean())/(df['MarkDown4'].std())
df['MarkDown5'] = (df['MarkDown5'] - df['MarkDown5'].mean())/(df['MarkDown5'].std())

# CPI standardize
df['CPI'] = (df['CPI'] - df['CPI'].mean())/(df['CPI'].std())

#Temperature Scaled
df['Temperature'] = (df['Temperature'] - df['Temperature'].mean())/(df['Temperature'].std())


#Fuel_Price Scaled
df['Fuel_Price'] = (df['Fuel_Price'] - df['Fuel_Price'].mean())/(df['Fuel_Price'].std())

# Size Scaled
df['Size'] = np.log(df['Size'])

df['Size'] = (df['Size'] - df['Size'].mean())/(df['Size'].std())
```
