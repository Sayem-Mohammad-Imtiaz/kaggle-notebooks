## Brief EDA & Visualisations
-------

```python
source = frame.groupby(['channelGrouping',
            'date'])['pageviews'].sum().reset_index()
source['date'] = pd.to_datetime(source['date'])

alt.Chart(source).mark_circle(
    opacity=0.8,
    stroke='black',
    strokeWidth=1
).encode(
    alt.X('monthdate(date):N', title='date',
          axis=alt.Axis(labelAngle=45)),
    alt.Y('channelGrouping:N'),
    alt.Size('pageviews:Q',
        scale=alt.Scale(range=[0, 1800]),
        legend=alt.Legend(title='No. of views')
    ),
    alt.Color('channelGrouping:N', legend=None)
).properties(
    title='Day-by-Day views Splitted by Channel',
    height=600
).transform_filter(
    alt.datum.Entity != ''
)
```

```python
source = frame.groupby(['browser',
            'date'])['pageviews'].sum().reset_index()
source['date'] = pd.to_datetime(source['date'])

chart = alt.Chart(source).mark_circle(
    opacity=0.8,
    stroke='black',
    strokeWidth=1
).encode(
    alt.X('monthdate(date):N', title='date',
          axis=alt.Axis(labelAngle=45)),
    alt.Y('browser:N'),
    alt.Size('pageviews:Q',
        scale=alt.Scale(range=[0, 1770]),
        legend=alt.Legend(title='No. of views')
    ),
    alt.Color('browser:N', legend=None)
).properties(
    title='Day-by-Day Views Splitted by Browser',
).transform_filter(
    alt.datum.Entity != ''
)

alt.layer(chart)
```

```python
source = frame.groupby('date')[
    ['visits','hits','pageviews']].sum().reset_index()
source['date'] = pd.to_datetime(source['date'])
source = source.reset_index().melt('date', 
    var_name='category', value_name='y')

nearest = alt.selection(type='single', nearest=True,
            on='mouseover', fields=['date'], empty='none')

cls_ = ['#001f3f','#FF4136','#0074D9']
domain = ['visits','hits','pageviews']

line = alt.Chart(source).mark_line(interpolate='monotone').encode(
    alt.X('monthdate(date):N', title='date',
          axis=alt.Axis(labelAngle=45)),
    alt.Y('y:Q',title='Number of Actions',
         axis=alt.Axis(gridOpacity=.5)),
    color=alt.Color('category:N',
        scale=alt.Scale(domain=domain,range=cls_)
                   ),
    opacity=alt.value(.75)
).properties(title='Day-by-day changing in number of visits, hits & pageviews')

selectors = alt.Chart(source).mark_point().encode(
    x='monthdate(date):N',
    opacity = alt.value(0)
).add_selection(nearest)

points = line.mark_point().encode(
    opacity=alt.condition(
        nearest,alt.value(1), alt.value(0))
)

text = line.mark_text(align='left',dx=5,dy=-13).encode(
    text=alt.condition(nearest,'y:Q', alt.value(''))
).transform_filter(nearest)

rule = alt.Chart(source).mark_rule(color='basis').encode(
    x='monthdate(date):N'
).transform_filter(nearest)

alt.layer(
    line, selectors, points, text, rule).properties(
        height=300, width=800)
```
