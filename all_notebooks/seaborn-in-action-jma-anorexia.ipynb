```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
import os 
import pandas as pd

dat = pd.read_csv('../input/anorexia-data-on-weight-change/anorexia.csv')
dat.head()
```

```python
sns.lmplot(x="Prewt", y="Postwt", hue="Treat",  data=dat);
```

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(dat.Prewt, dat.Postwt, ax=ax)
sns.rugplot(dat.Prewt, color="g", ax=ax)
sns.rugplot(dat.Postwt, vertical=True, ax=ax);
```

```python
f, ax = plt.subplots(figsize=(6, 6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(dat.Prewt, dat.Postwt, cmap=cmap, n_levels=60, shade=True);
```

```python
g = sns.jointplot(x="Prewt", y="Postwt", data=dat, kind="kde", color="m")
g.plot_joint(plt.scatter, c="w", s=30, linewidth=1, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$Prewt", "$Postwt")
```
