This is a dataset about Wage and classification information of City of Seattle Employees as of January 2020. According to City of Seattle, exclusions may apply for vulnerable populations.

The data is downloaded from Seattle data portal below:
https://data.seattle.gov/City-Business/City-of-Seattle-Wage-Data/2khk-5ukd/data

First of all, I would like to import the csv file and get familarize with the dataset. 
1) Check if there is any missing ( null ) value in the set
2) Check the last 5 rows of dataset 



```python
import matplotlib.pyplot as plt
import seaborn as sns

#read data into dataframe named 'wage'
import pandas as pd
wage = pd.read_csv("../input/City_of_Seattle_Wage_Data.csv")

# Check if there is missing value in dataset
wage.isnull().sum()

```

```python
#check the last 5 rows in dataset
wage.tail()
```

Under the Job title column, some job title ends with an asterisk.
I have decided to do some clean up and take out the asterisk. Also, for column label " Hourly Rate ", there is an extra leading space to be cleaned up. 

```python
#remove * and leading space 
wage['Job Title'] = wage['Job Title'].str.replace('*','').str.strip()

#remove leading space in column label - 'Hourly Rate '
wage.rename(columns={'Hourly Rate ':'Hourly Rate'}, inplace=True)

#view update
wage.tail()
```

Let me find out how many data entries are there per Department.

```python
wage['Department'].value_counts()
```

Let me create a box plot to analyse the hourly rate distribution per department.

```python
f, ax = plt.subplots(figsize=(8, 15))
sns.boxplot(x='Hourly Rate', y='Department', data=wage)
plt.title('Distribution of Hourly Rates')
```

Now, I would like to look into 'Hourly Rate'. What are the max, min and mean ?

```python
wage['Hourly Rate'].describe()
```

```python
hist_plot=wage.hist(column='Hourly Rate', bins=100, figsize=(8,8))
plt.title('Histogram of Hourly Rate among City of Seattle Employees in Jan 2020')
plt.xlabel('Hourly Rate')
plt.ylabel('Frequency')
```

From above, I am surprised to find out the min. hourly rate in the dataset is 5.53, which is below min. wage of $16.39 in City of Seattle. Let us find out how many entries in this dataset are below min. wage and their corresponding job title per department. 

```python
def underpaid(rate):
    if rate < 16.39:
         return ("Below Min. Wage")
    elif rate >= 16.39:
        return ("Meet Min. Wage")
```

```python
wage['Hourly Rate'].apply(underpaid).value_counts()
```

There are 6 entries of "Below Min. Wage". Who are they?

```python
wage[wage['Hourly Rate']<16.39]
```

****Conclusion****: Civil Svc Commissioner in Civil Service Commissions are the least earners in this dataset with an hourly rate of $5.53, which is way before min. wage in City of Seattle in 2020.

Now, let's look at the top 10 most earners in the dataset. 

```python
wage.nlargest(10, ['Hourly Rate'])
```

**Findings**: The highest earner in the dataset is City Light General Mgr & CEO in Seattle City Light department and has an hourly rate of $175 !
