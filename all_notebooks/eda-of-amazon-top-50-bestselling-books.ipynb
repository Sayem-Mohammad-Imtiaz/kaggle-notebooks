```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/kaggle/input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv")
```

```python
#dummy variable allocation for genre
dummies=pd.get_dummies(df["Genre"])
df=pd.concat([df,dummies],axis=1)
df
```

```python
df=df.drop(["Genre"],axis=1)
#here we can easly remove both genre and non fiction dummy variable created as from "Fiction " variable itself we could get the information about non fiction variable
```

```python
df.head()
```

```python
sns.pairplot(df,diag_kind="kde",kind='scatter')
```

```python
#correlation heatmap visualization
corr=df.corr()
sns.heatmap(corr)
```

# 1.Which book has maximum and minimum price


```python

set(["{} is the book with maximum price: {}".format(i,j) for i,j in zip(df["Name"],df["Price"]) if j==df["Price"].max()])
```

```python
set(["{} is the book with minimum price: {}".format(i,j) for i,j in zip(df["Name"],df["Price"]) if j==df["Price"].min()])
```

# Grouping by author

```python
df["Author"].nunique()
```

```python
group=df.groupby("Author")
group
```

```python
group.sum()
#neglect Year as it gives the sum of multiple years in the same group
```

Answer for some questions

# 1.which author has maximum number of fictions sold

```python

group.sum().sort_values("Fiction").tail()
#from this output we can see that JEFF KINNEY -12 fiction book were sold out and this topped the list of max fictional book sold
```

# 2.which author has maximum number of non fiction sold

```python

group.sum().sort_values("Non Fiction").tail()
#Gary Chapman's 11 non fictional book were sold
```

# 3.sum of prices of all the books of a particular author being sold

```python

group.sum().sort_values("Price").tail()
# here the total max price is of American Psychological Association ie 460
```
