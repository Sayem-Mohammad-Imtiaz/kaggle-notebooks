Inspite of the dangers of Nuclears energy, the LCOE associated with Nuclear energy is low as compared to other renewbale and non renewable sources .
At a steady O&M cost , nuclear forms the base load of any producer . 
We can see gthe carbon price and heat proce at 30 and 35 USD/$MWh

```python
# Max values 
plot_nuclear= df_nuclear[df_nuclear["Total capital costs (USD/MWh)"]==df_nuclear["Total capital costs (USD/MWh)"].max()].plot.bar()
plot_nuclear.set_title('LCOE Nuclear')
plot_nuclear.set_xlabel("Components")
plot_nuclear.set_ylabel("Cost (USD/MWh)")
plot_nuclear.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

## Renewables

### Wind LCOE

```python
mean = df_wind.mean().round(2)
mean
```

```python
plot_wind = mean.plot.bar()
plot_wind.set_title('LCOE Wind')
plot_wind.set_xlabel("Components")
plot_wind.set_ylabel("Cost (USD/MWh)")
```

Wind is a renewable source driven by natural wind and hence fuel cost is very low ( 10 USD per MWh ) which could be attributed to electrical compoennets in the system. 
Unknown heat proce in Wind's LCOE data needs to be investigated 

```python
# df_wind_max = df_wind[df_wind["Total capital costs (USD/MWh)"]==df_wind["Total capital costs (USD/MWh)"].max()]
# df_wind_max.head()
```

```python
df_wind_median = df_wind[df_wind["Plant type"].astype(str).str.contains("median case")]
df_wind_median = df_wind_median[df_wind_median["Total capital costs (USD/MWh)"]==df_wind_median["Total capital costs (USD/MWh)"].max()]
df_wind_median.head()
```

```python
plot_wind = df_wind_median.plot.bar(legend=False)
plot_wind.set_title('LCOE wind')
plot_wind.set_xlabel("Components")
plot_wind.set_ylabel("Cost (USD/MWh)")
# plot_wind.xaxis.set_tick_params(labelbottom=True)
# plot_wind.yaxis.set_tick_params(labelleft=True)
plot_wind.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
df_wind_median['LCOE (cents/kWh)'] = df_wind_median['LCOE (USD/MWh)']*100/1000
df_wind_median[['Plant type','LCOE (cents/kWh)','LCOE (USD/MWh)']].transpose()
```

## Non renewable

```python
cols = ['Construction costs (USD/MWh)','Refurbishment costs (USD/MWh)','Decommissioning costs (USD/MWh)','Total capital costs (USD/MWh)',
            'Operations & maintenance costs (USD/MWh)','Fuel costs (th) (USD/MWh)','Fuel costs (el) (USD/MWh)','Carbon price (USD/ton)',
            'Carbon costs (USD/MWh)','Heat price (USD/MWh)','CHP heat revenues (USD/MWh)','Discount rate','LCOE (USD/MWh)']
    
def addlabels(plt , df):
    for i, v in enumerate(cols):
    #     print(i + " " + v + ' ' + str(df_coal_median[v].values[0]))
        plt.text(i*0.04 - 0.24,2, 
                       v + ' ' + str(df[v].values[0]),
                rotation=90)
```
