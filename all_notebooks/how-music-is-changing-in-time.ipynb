```python
data = pd.read_csv('/kaggle/input/discogs-database-all-release-data/release_data/release_data.csv')
```

```python
def counts_bar(*args):
    "count total formats, genre and countries and plot bar plot"
    global genre_counts, format_counts, country_counts
    
    data_pivot = data.reset_index()
    data_pivot = data_pivot[data_pivot!='Europe']

    for arg in args:
          
            var_counts = data_pivot.pivot_table(index=arg, values='release_id',aggfunc=lambda x: len(x.unique())).\
            sort_values(by='release_id',ascending=False)

            if arg == 'genre':
                genre_counts = var_counts.head(8)
                genre_counts = genre_counts.rename({'Folk, World, & Country':'Folk/Country'})
            elif arg == 'format':
                format_counts = var_counts.head(4)
            elif arg == 'country':
                country_counts = var_counts.head(8)
                
            del var_counts
                
    fig, axes =  plt.subplots(1,3,figsize=(33,7))
    sns.set(style='darkgrid')
    sns.despine()
    
    genre_plot=sns.barplot(x=genre_counts.index, y=genre_counts['release_id'].div(10**6), data=genre_counts,ax=axes[0])
    genre_plot.set_ylabel('number of releases (million)',fontsize=20)
    format_plot = sns.barplot(x=format_counts.index, y=format_counts['release_id'].div(10**6), data=format_counts,ax=axes[1])
    country_plot=sns.barplot(x=country_counts.index, y=country_counts['release_id'].div(10**6), data=country_counts,ax=axes[2])

    for i,k in enumerate(args):
        axes[i].tick_params(axis='both',which='major',labelsize=20)
        axes[i].tick_params(axis='x',which='major',rotation=45)
        eval(k+'_plot').set_ylabel('')
        eval(k+'_plot').set_xlabel(k,fontsize=20)
        axes[i].set_title('total number of releases by '+k,{'fontsize':20},pad=15)
        
    genre_plot.set_ylabel('number of releases (million)',fontsize=20)
    plt.show()
    plt.close(fig)
  
    return []

var1 = 'genre'
var2 = 'format'
var3 = 'country'
counts_bar(var1, var2, var3)
```

```python
def year_plots(*args):
    global genre_year, format_year, country_year
    data_pivot = data.reset_index()
    data_pivot = data_pivot[data_pivot!='Europe']
    
    for arg in args:
        
            var_year = data_pivot.pivot_table(index=arg,columns='year', values='release_id',aggfunc=lambda x: len(x.unique()))

            if arg == 'genre':
                genre_year = var_year
                genre_year = genre_year.rename({'Folk, World, & Country':'Folk/Country'})
                genre_year = genre_year[genre_year.index.isin(genre_counts.index)].transpose()
                
                
            elif arg == 'format':
                format_year = var_year
                format_year = format_year[format_year.index.isin(format_counts.index)].transpose()
                
            elif arg == 'country':
                country_year = var_year  
                country_year = country_year[country_year.index.isin(country_counts.index)].transpose()
   
    fig, axes = plt.subplots(3,1,figsize=(12,20))
    sns.set(style='darkgrid')
    
    genre_year_plot=sns.lineplot(data=genre_year,ax=axes[0],dashes=False)
    format_year_plot=sns.lineplot(data=format_year,ax=axes[1],dashes=False)
    country_year_plot=sns.lineplot(data=country_year,ax=axes[2],dashes=False)
    
    for i,k in enumerate(args):
        
        axes[i].set_xlim(left=1940)
        axes[i].set_ylabel('number of releases',fontsize=20)
        axes[i].tick_params(axis='both',which='major',labelsize = 20)
        axes[i].set_xlabel('year',fontsize=20)
        axes[i].legend(fontsize='x-large',fancybox=True,shadow=True,facecolor='white')
        axes[i].set_title(k + ' releases by year',{'fontsize':20},pad=15)
    
    plt.subplots_adjust(hspace=0.34)
    plt.show()

    return []

var1 = 'genre'
var2 = 'format'
var3 = 'country'
year_plots(var1, var2, var3)
```

```python
#what has been happening in the last two decades?
format_year_2000 = format_year[(format_year.index>=2000) & (format_year.index<=2020)]

fig,axes = plt.subplots(figsize=(15,9))
sns.set(style='darkgrid')  
format_year_2000_plot=sns.lineplot(data=format_year_2000,ax=axes,dashes=False)

axes.set_ylabel('number of releases',fontsize=20)
axes.tick_params(axis='both',which='major',labelsize = 20)
axes.set_xlabel('year',fontsize=20)
axes.legend(fontsize='x-large',fancybox=True,shadow=True,facecolor='white')
axes.set_title('formats in the 2000s',{'fontsize':20},pad=15)

plt.show()
```
