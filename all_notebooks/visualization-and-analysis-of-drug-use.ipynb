```python
file_path = "../input/drug-use-by-age/drug-use-by-age.csv"
Original_Data = pd.read_csv(file_path)
Original_Data
```

In this dataset NaN values have "-"(Hyphen),  
 So we are going to replace "-" with "0"

```python
Data = Original_Data.replace('-',0)
Data
```

Now we will check for missing values

```python
Data.isnull().sum()
```

Fantastic no missing values. 
 Now let us visualize the data

```python
Data.head()
```

```python
plt.figure(figsize= (16,6))
sns.lineplot(x="age", y = "n",data = Data)
```

```python
plt.figure(figsize=(16,6))
plt.title("Age and number")
sns.barplot(x=Data['age'], y = Data['n'],color ="grey")
```

As we can see from the graph above,
the highest number of drug consumers are in the age group of 35-49

```python
species_use = ["alcohol-use", "marijuana-use","cocaine-use","crack-use","heroin-use","hallucinogen-use","inhalant-use","pain-releiver-use","oxycontin-use","stimulant-use","meth-use","sedative-use"]
drugs_use = Data[species_use]
drugs_use.info   
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(x='age',y = "alcohol-use", data= Data, label = "alcohol-use")
sns.lineplot(x='age',y = "marijuana-use", data= Data, label = "marijuana")
sns.lineplot(x='age',y = "cocaine-use", data= Data, label = "cocaine")
sns.lineplot(x='age',y = "crack-use", data= Data, label = "crack")
sns.lineplot(x='age',y = "heroin-use", data= Data, label = "heroin")
sns.lineplot(x='age',y = "hallucinogen-use", data= Data, label = "hallucinogen")
sns.lineplot(x='age',y = "inhalant-use", data= Data, label = "inhalant")
sns.lineplot(x='age',y = "pain-releiver-use", data= Data, label = "pain-releiver")
sns.lineplot(x='age',y = "oxycontin-use", data =Data, label = "oxycontin")
sns.lineplot(x='age',y = "stimulant-use", data =Data, label = "stimulant")
sns.lineplot(x='age',y = "meth-use", data =Data, label ="meth")
sns.lineplot(x='age',y = "sedative-use", data =Data, label ="sedative")
```

```python
species_freq = ["alcohol-frequency", "marijuana-frequency","cocaine-frequency","crack-frequency","heroin-frequency","hallucinogen-frequency","inhalant-frequency","pain-releiver-frequency","oxycontin-frequency","stimulant-frequency","meth-frequency","sedative-frequency"]
drugs_freq = Data[species_freq]
drugs_freq.info 
```
