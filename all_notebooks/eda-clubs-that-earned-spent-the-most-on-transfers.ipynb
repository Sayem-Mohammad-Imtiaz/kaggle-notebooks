```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ftbl1 = ftbl.groupby('From(Club)', as_index=False,sort=True)['Fee(€ mln)'].sum()
ftbl1 = ftbl1.sort_values('Fee(€ mln)', ascending=False)
ftbl1.head()
```

```python
ftbl1 = ftbl1[0:10]
plt.figure(figsize=(15,12))
sns.set_style('darkgrid')
p = sns.barplot(x='From(Club)',y='Fee(€ mln)',data=ftbl1)
p.set( xlabel = "Name of the club", ylabel = "sum of money earned in transfers")
p.set_title("10 clubs with the highest transfer revenues")
```

```python
ftbl2 = ftbl.groupby('To(Club)', as_index=False,sort=True)['Fee(€ mln)'].sum()
ftbl2 = ftbl2.sort_values('Fee(€ mln)', ascending=False)
ftbl2.head()
```

```python
ftbl2 = ftbl2[0:10]
plt.figure(figsize=(18,15))
sns.set_style('darkgrid')
p = sns.barplot(x='To(Club)',y='Fee(€ mln)',data=ftbl2)
p.set( xlabel = "Name of the club", ylabel = "sum of money spent in transfers")
p.set_title("top 10 clubs that spent the most on footballers")
```

```python
ftbl3 = ftbl.groupby('Origin', as_index=False,sort=True)['Fee(€ mln)'].sum()
ftbl3 = ftbl3.sort_values('Fee(€ mln)', ascending=False)
ftbl3.head()
```

```python
ftbl3 = ftbl3[0:5]
plt.figure(figsize=(18,15))
sns.set_style('darkgrid')
p = sns.barplot(x='Origin',y='Fee(€ mln)',data=ftbl3)
p.set( xlabel = "Name of the Country", ylabel = "sum of money")
p.set_title("top 5 countries where the most expensive players come from")
```

# Let's check the average price of a player depending on his position

```python
ftbl4 = ftbl.groupby('Position', as_index=False,sort=True)['Fee(€ mln)'].sum()
ftbl4 = ftbl4.sort_values('Fee(€ mln)', ascending=False)
ftbl4.head()
```

```python
ftbl5 = ftbl['Position'].value_counts()
ftbl5 = pd.DataFrame(ftbl5)
ftbl5.columns =[ 'Count']
ftbl5
```

```python
ftbl4['Count'] = ftbl5['Count'].values
ftbl4['Average_price'] = ftbl4['Fee(€ mln)']/ftbl4['Count']
ftbl4
```

# Chart instruction: percentage% (average price in € mln)

```python
import matplotlib.pyplot as plt
def make_autopct(values):
    def my_autopct(pct):
        total = sum(values)
        val = int(round(pct*total/100.0))
        return '{p:.2f}%  ({v:d})'.format(p=pct,v=val)
    return my_autopct
mlabels=['Midfielder','Forward','Striker','Defender','Goalkeeper']
plt.pie(ftbl4['Average_price'],labels=mlabels,autopct=make_autopct(ftbl4['Average_price']))
plt.title('Comparison of average earnings by position of the player')
fig = plt.gcf()
fig.set_size_inches(12,12)
plt.show()
```
