```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, Conv2D 
from tensorflow.keras.layers import MaxPooling2D
```

First we load our data and split them into train and test sets.
Our label column which is our target column, has 10 values, we convert this column into categorical data where each number represents as follow:

* 0 T-shirt/top
* 1 Trouser
* 2 Pullover
* 3 Dress
* 4 Coat
* 5 Sandal
* 6 Shirt
* 7 Sneaker
* 8 Bag
* 9 Ankle boot

The last thing to prepare our dataset for modeling is to reshape our training data images into 28x28 pixel images.

```python
train_data = np.loadtxt("../input/fashion-mnist_train.csv", delimiter=',',
                       skiprows=1)
x_train = train_data[:,1:]
y_train = train_data[:,0]

test_data = np.loadtxt("../input/fashion-mnist_test.csv", delimiter=',',
                      skiprows=1)
x_test = test_data[:,1:]
y_test = test_data[:,0]

y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)

x_train = x_train.reshape(x_train.shape[0], 28,28,1)/225
x_test = x_test.reshape(x_test.shape[0], 28,28,1)/225
```

Let's construct our model with the layers.

```python
model = Sequential()
model.add(Conv2D(32,(3,3), padding="same", activation="relu", 
                 input_shape=(28,28, 1)))
model.add(Conv2D(32,(3,3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Conv2D(64,(3,3), padding="same", activation="relu"))
model.add(Conv2D(64,(3,3), activation="relu"))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))

model.add(Flatten())

model.add(Dense(512, activation="relu"))
model.add(Dropout(0.5)) 
model.add(Dense(10, activation="softmax"))

model.compile(loss="categorical_crossentropy", optimizer="adam", 
              metrics=["accuracy"])
```

```python
model.fit(x_train, y_train, batch_size=32, epochs=10,
          validation_data=(x_test, y_test), shuffle=True)
```

We have approximately 93% accuracy for our model which is very good.

```python
model.evaluate(x_test, y_test, batch_size=100)
```
