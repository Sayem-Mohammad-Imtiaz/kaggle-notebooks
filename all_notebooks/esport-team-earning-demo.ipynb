```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# I'm using plotly because it's simple and elegance
import plotly.express as px
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objects as go
init_notebook_mode(connected=True)

# So that warning doesn't pop up
import warnings
warnings.filterwarnings("ignore")
```

```python
team = pd.read_csv('/kaggle/input/top-100-highest-overall-esport-team-earning/ESPORT Team Dataset/esport_team.csv')
country = pd.read_csv('/kaggle/input/top-100-highest-overall-esport-team-earning/ESPORT Team Dataset/esport_country.csv')
player = pd.read_csv('/kaggle/input/top-100-highest-overall-esport-team-earning/ESPORT Team Dataset/esport_player.csv')
game = pd.read_csv('/kaggle/input/top-100-highest-overall-esport-team-earning/ESPORT Team Dataset/competitive_game.csv')
```

```python
# Checking Missing value
for df in [team,country,player,game]:
    print(df.isna().sum())
    print("--"*20)
```

```python
for df in [team,country,player,game]:
    print(df.columns.tolist())
```

```python
# Converting values that is in US dollar into floating point value

def convertToFloat(string):
    '''Converting specific string pattern to floating point value'''
    try:
        result = float(string.replace("$","").replace(",",""))
    except:
        result = np.nan
    return result

# Converting strign percentage value to integer
def convertToNum(string):
    '''Convert percent (string) to integer'''
    try:
        result = float(string.replace("%",""))
    except:
        result = np.nan
    return result

# Converting
team['Total Prize Won'] = team['Total Prize Won'].apply(lambda s: convertToFloat(s))
country['Total Earning by Player'] = country['Total Earning by Player'].apply(lambda s: convertToFloat(s))
country['Percentage of Country Total Earning'] = country['Percentage of Country Total Earning'].apply(lambda s: convertToNum(s))
player['Total Earning'] = player['Total Earning'].apply(lambda s: convertToFloat(s))
game['Total Money Earned'] = game['Total Money Earned'].apply(lambda s: convertToFloat(s))
game['Total Money Earned Percentage'] = game['Total Money Earned Percentage'].apply(lambda s: convertToNum(s))
```

```python
# Checking for NaN value after running the convertsion function
country[country['Total Earning by Player'].isna()]
```

```python
# Dropping the NaN value
for df in [team,country,player,game]:
    df.dropna(inplace=True)
```

```python
# Checking for the Invalid again after dropping NaN values
country[country['Total Earning by Player'].isna()]
```

```python
# Create new column for all dataset that convert Total Prize Won into billion so easier analysis

team['Total Prize (Million)'] = team['Total Prize Won'].apply(lambda prize: prize / 10**6)
country['Total Earning by Player (Million)'] = country['Total Earning by Player'].apply(lambda prize: prize / 10**6)
```

```python
# Exploring the Data

# Team with highger price won than the mean

# team[team['Total Prize Won'] >= team['Total Prize Won'].mean()]
# team[team['Total Prize Won'] > team['Total Prize Won'].quantile(0.75)]

team.sort_values(by='Total Prize Won',ascending = False).head()
```

```python
df = country.groupby('Country').sum().sort_values(by='Total Earning by Player',ascending = False)

fig = px.bar(data_frame = df.head(15),
             x = df.head(15).index,
             y = 'Total Earning by Player')

fig.update_layout(title = 'Top 10 Countries by Total Player Earning')

fig.update_xaxes(tickangle=45,
                 title = 'Country')

fig.show()
```

```python
# Highest Cummulative Price given by a game
df = game.groupby('Game').sum().sort_values('Total Money Earned',ascending = False).head(10)

fig = px.bar(data_frame = df,
             x = df.index,
             y = 'Total Money Earned')

fig.show()
```
