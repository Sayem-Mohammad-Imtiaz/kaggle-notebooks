# Genre preference

```python
# Peopels preferred genre
books.Genre.value_counts().plot.barh()
```

# Authors preference

```python
# Preferred author
books.Author.value_counts().head(20).plot.barh()
# 248 authors lets find top 20
```

```python
# Tagging authors based on their Genre
Top_authors=list(books.Author.value_counts().head(20).index)
Genre_author=dict()
for author in Top_authors:
    Genre_author[author]=books[books.Author=="Jeff Kinney"]["Genre"].mode()[0]
Genre_author    
```

# Price preference

```python
# maximum cost of book sold
print("max:",books.Price.max(),"min:",books.Price.min(),"mode:",books.Price.mode()[0])

# maximum cost of top 20 authors
print("max:",books[books.Author.isin(Top_authors)].Price.max(),"min:",books[books.Author.isin(Top_authors)].Price.min(),"mode:",books[books.Author.isin(Top_authors)].Price.mode()[0])

books[books.Price<=80].Price.plot.hist(bins=50)
books[books.Author.isin(Top_authors)].Price.plot.hist(bins=50)
```

# Ratings

```python
print("max:",books.Reviews.max(),"min:",books.Reviews.min(),"mode:",books.Reviews.mode()[0])

# maximum cost of top 20 authors
print("max:",books[books.Author.isin(Top_authors)].Reviews.max(),"min:",books[books.Author.isin(Top_authors)].Reviews.min(),"mode:",books[books.Author.isin(Top_authors)].Reviews.mode()[0])

books[books.Reviews<40000].Reviews.plot.hist(bins=50)
books[books.Author.isin(Top_authors)].Reviews.plot.hist(bins=50)
```
