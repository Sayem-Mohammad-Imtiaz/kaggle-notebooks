Importing Libraries 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Importing DataSet

```python
dataset = pd.read_csv('../input/second-hand-used-cars-data-set-linear-regression/train.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
y=y.reshape(len(y),1)
```

Splitting data into training and testing set 

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
sc_y=StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test=sc_X.fit_transform(X_test)
y_train=sc_y.fit_transform(y_train)
```

Training the Linear Regression on training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

Predicting values for X_test

```python
y_pred=sc_y.inverse_transform(regressor.predict(X_test))
```

Printing predicted values and test values of y in two column of same matrics

```python
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

You can see from above my model working fine on test values Now finding the R square values

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

***R Sqaure value is very near to 1 so can say This Model fits good for this data set***
