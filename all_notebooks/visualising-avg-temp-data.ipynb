# Visualising Temperature Data
In this we will clean and analyse the dataset. Then we will try to plot the average temperature of all the countries.

```python
# importing required packages
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.express as px
from plotly.offline import init_notebook_mode
init_notebook_mode(connected=True)
```

```python
# reading the data
global_temp_country=pd.read_csv('/kaggle/input/global-land-temperatures-by-country/GlobalLandTemperaturesByCountry.csv')
global_temp_country.head()
```

```python
global_temp_country.shape
```

```python
global_temp_country.isna().sum()
```

We can see that there are many missing values. We will now drop whole rows where any one column has a missing value.

```python
global_temp_country.dropna(axis = 'index', how = 'any', subset = ['AverageTemperature'], inplace = True)
```

```python
global_temp_country['Country'].nunique()
```

```python
global_temp_country['Country'].unique()
```

We see that there are some duplications and we have to remove that.

```python
dict={
    'Denmark (Europe)':'Denmark',
    'Congo (Democratic Republic Of The)':'Congo',
    'France (Europe)':'France',
    'Netherlands (Europe)':'Netherlands',
    'United Kingdom (Europe)':'United Kingdom'
}
```

```python
global_temp_country['Country'].replace(dict,inplace=True)
```

#### Analyse Average Temperature of Each Country

```python
avg_temp=global_temp_country.groupby(['Country'])['AverageTemperature'].mean().to_frame().reset_index()
avg_temp.head()
```

#### Choropleth Map

```python
fig=px.choropleth(avg_temp, locations='Country', locationmode='country names', color='AverageTemperature')
fig.update_layout(title='Choropleth Map of Average Temperature')
fig.show()
```
