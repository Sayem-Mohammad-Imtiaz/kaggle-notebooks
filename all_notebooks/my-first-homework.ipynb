**Reading .csv file with " pd.read_csv() ". Then the datas which is read by pd.read_csv() is stored in the " data " variable.**

**The heatmap shows the relationship between features of dataset. **

```python
#Correlation Map

f,ax = plt.subplots(figsize=(18,18))
sns.heatmap(data.corr(), annot= True, linewidths= .5, fmt='.1f', ax= ax)
plt.show()
```

**Plotting simple line graph of the Job Level and Total Working Years.**

```python
data.Age.plot( kind='line', color='blue', label='Age', linewidth=1, alpha=.5, grid= True, linestyle=':')
data.TotalWorkingYears.plot(kind='line', color= 'green', label= 'Total Working Years', linewidth=1, alpha=.5, grid= True, linestyle='-.')

plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('Number of Employees')              # label = name of label
plt.ylabel('Years')
plt.title('Line Plot of the Years-Employees Relation')            # title = title of plot
plt.show()
```

**Plotting simple scatter graph of the Job Level and Total Working Years.**

```python
# Scatter Plot 
# x = Age, y = Total Working Years
data.plot(kind='scatter', x='Age', y='TotalWorkingYears',alpha = 0.5,color = 'red')
plt.xlabel('Age')              # label = name of label
plt.ylabel('Total Working Years')
plt.title('Age-Total Working Years Scatter Plot')            # title = title of plot
plt.show()
```

**Plotting simple histogram of the employees' ages.**

```python
data.Age.plot(kind='hist', color = 'MediumSeaGreen', bins=60, figsize = (20,12))
plt.show()
```

**Clearing the histogram of the age.**

**Employees which are bigger than 59 years old are chosen.**

```python
x = data['Age'] > 59
data[x]
```

**Employees which are married and bigger than 59 years old are chosen.**

```python
data[np.logical_and(data['Age']>59, data['MaritalStatus'] == 'Married' )]
```

**The above code can be written differently, such as the following codes.**

```python
# First Type
data[np.logical_and(data['Age'] > 59, data['MaritalStatus'] == 'Married')]
```

```python
# Second Type
data[(data['Age']> 59) & (data['MaritalStatus'] == 'Married')]
```
