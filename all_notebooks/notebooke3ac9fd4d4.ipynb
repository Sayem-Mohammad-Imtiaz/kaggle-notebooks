```python
import numba as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

```python
df=pd.read_csv("../input/fifa-world-cup-match-stats/FIFAallMatchBoxData.csv")
```

```python
plt.figure(figsize=(20,10))
sns.countplot(df['hname'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(23,10))
plt.style.use("fivethirtyeight")
sns.lineplot(x='hname',y='hgoals',data=df)
plt.xticks(rotation=90)
plt.show()
```

```python
df['hname'].value_counts().head(10)
```

```python
df5=df[(df["hname"]=="Argentina") |(df["hname"]=="Brazil")|(df["hname"]=='Uruguay') |(df["hname"]=='Germany')|(df["hname"]=='Colombia')]
```

```python
plt.figure(figsize=(20,7))
plt.style.use('ggplot')
sns.lineplot(data=df5,x="year",y="hyellowCards",marker='o',hue="hname")
plt.show()
```

```python
sns.pairplot(data=df)
```

```python
plt.figure(figsize=(23,10))
plt.style.use("fivethirtyeight")
sns.lineplot(x='year',y='hgoals',data=df, label='hgoals')
sns.lineplot(x='year',y='agoals',data=df, label='agoals')

sns.lineplot(x='year',y='hshotsOnTarget',data=df, label='hshotsOnTarget')
sns.lineplot(x='year',y='ashotsOnTarget',data=df, label='ashotsOnTarget')

plt.legend()

plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(23,10))
sns.lineplot(x='year',y='hPossesion',data=df, label='hPossesion')
sns.lineplot(x='year',y='aPossesion',data=df, label='aPossesion')
```

```python
plt.figure(figsize=(23,10))
plt.style.use("fivethirtyeight")
sns.lineplot(x='year',y='hyellowCards',data=df, label='hyellowCards')
sns.lineplot(x='year',y='ayellowCards',data=df, label='ayellowCards')

sns.lineplot(x='year',y='hredCards',data=df, label='hredCards')
sns.lineplot(x='year',y='aredCards',data=df, label='aredCards')

plt.legend()

plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(23,10))
plt.style.use("fivethirtyeight")
sns.lineplot(x='year',y='hfouls',data=df, label='hfouls')
sns.lineplot(x='year',y='afouls',data=df, label='afouls')

sns.lineplot(x='year',y='hsaves',data=df, label='hsaves')
sns.lineplot(x='year',y='asaves',data=df, label='asaves')

plt.legend()

plt.xticks(rotation=90)
plt.show()
```
