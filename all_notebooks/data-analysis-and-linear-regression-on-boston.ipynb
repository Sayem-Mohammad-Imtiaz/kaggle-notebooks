```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.DESCR)
boston_df = boston.data
```

```python
#Creating df
df=pd.DataFrame(boston.data,columns=boston.feature_names)
```

```python
#Adding the target to the df
df['TARGET']=boston.target
```

```python
#Getting information about the dataset
df.info()
```

```python
df.head()
```

```python
sns.pairplot(df[['CRIM', 'ZN', 'INDUS', 'NOX', 'TARGET']])
```

**It is visible that as the crime rate(CRIM) increases the value of the house (TARGET) decreases.**

**As NOX value increases the Target value decreases**

**The relation between NOX value and INDUS (Industries i guess) is also visible.**

```python
sns.pairplot(df[['RM', 'AGE', 'DIS', 'RAD', 'TARGET',]])
```

**As no of rooms(RM) increases the value of the house(TARGET) increases**


```python
sns.pairplot(df[['TAX', 'PTRATIO', 'B', 'LSTAT', 'TARGET']])
```

**Relation between LSTAT and TARGET is also seen**

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['TARGET'],axis=1)
y=df['TARGET']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
lm.coef_
```

```python
pred = lm.predict(X_test)
```

```python
sns.scatterplot(x=y_test,y=pred)
```

```python
#Accuracy
lm.score(X_test,y_test)
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
print('MAE:', mean_absolute_error(y_test,pred))
print('MSE:', mean_squared_error(y_test,pred))
print('RMSE:', np.sqrt(mean_squared_error(y_test,pred)))
```

```python
pd.DataFrame(lm.coef_,X.columns, columns=['Coeffecient'])
```

**Holding all other features fixed, a 1 unit increase in 'Crime rate (CRIM)' is associated with an decrease of -0.088505 target value(money).**

**Holding all other features fixed, a 1 unit increase in 'Nitric oxides concentration (NOX)' is associated with an decrease of -17.748371 target value(money).**

**Holding all other features fixed, a 1 unit increase in 'Average number of rooms per dwelling (RM)' is associated with an increase of 3.247765 target value(money).**
