SnapChat is one of the most famous social networking apps with more than 1,000,000,000+ installs, its ratings currently stands at 4.3 on Google Play. with 1 being the lowest and 5 being the highest. This is a quick analysis on the user ratings to understand the most commmon areas of interest by the the SnapChat users.

Objectives: 
1. To analyse the numerical ratings of SnapChat. 
2. To analyse the most common topics and keywords used in the user reviews. 

```python
sc_data = pd.read_csv('/kaggle/input/10k-snapchat-reviews/Snapchat_app_store_reviews.csv')
sc_data.reset_index(drop=True)
sc_data.head()
```

So now we havev a glimpse of the data, lets begin with analysing the data. 

```python
sc_data.info()
```

```python
# Converting the data to a sql query for easier data handling
sqldf('Select * from sc_data')
```

The average rating between 1 and 5 would be 3. So let's compare the count of ratings below average and above average. 

```python
Below_average = sqldf('Select COUNT(rating) as Less_than_3 from sc_data where rating < 3')
Above_average = sqldf('Select COUNT(rating) as More_than_3 from sc_data where rating >= 3')
print('Below Average {} \n Above Average {}'.format(Below_average,Above_average))
```

```python
Below_average = sqldf('Select rating as Less_than_3 from sc_data where rating < 3')
Above_average = sqldf('Select rating as More_than_3 from sc_data where rating >= 3')
labels1 = ['Below average ','Above Average']
value1 = [len(Below_average),len(Above_average)]

fig = go.Figure([go.Bar(x=labels1, y=value1)])
fig.show()
```

As we can see from the above graph, its an indication of good ratings for SnapChat as the majority of the ratings are above average. 

Next, lets analyse the overall number of ratings on SnapChat. 

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})

sns.distplot(a=sc_data['rating'], kde=True,rug=True,hist=True,color = 'g')
# sns.distplot(a=sc_data['rating'], label="1", kde=False)
plt.legend()

```

Insights from the above graph:
1. On a scale of user satisfaction, the majority of users gave it a four-star rating.
2. SnapChat's two-star rating has the least density.
