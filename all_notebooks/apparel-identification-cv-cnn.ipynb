```python
import cv2
```

```python
path_train = '/kaggle/input/identify-the-apparels-computer-vision/train_LbELtWX/train/' 
path_test = '/kaggle/input/identify-the-apparels-computer-vision/test_ScVgIM0/test/' 
```

```python
X_train=[]
for i in range(0,60000):
        i=i+1
        im = cv2.imread(path_train+'/'+str(i)+".png")
        im = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
        X_train.append(im)
```

```python
X_test=[]
for i in range(60000,70000):
        i=i+1
        im = cv2.imread(path_test+'/'+str(i)+".png")
        im = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
        X_test.append(im)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
X_train = np.array(X_train) 
X_train.shape
```

```python
X_test = np.array(X_test) 
X_test.shape
```

```python
y_train = pd.read_csv('/kaggle/input/identify-the-apparels-computer-vision/train_LbELtWX/train.csv')
y_train.head()
```

```python
y_train = y_train['label']
y_train = np.array(y_train)
```

```python
y_train.shape
```

```python
y_train[0:5]
```

```python
classes = ['T-shirt/top','Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
classes
```

```python
plt.imshow(X_train[0])
```

```python
# defining function to display images

def plot_sample(X, y, index):
    plt.figure(figsize=(15,2))
    plt.imshow(X[index])
    plt.xlabel(classes[y[index]])
```

```python
for i in range(0,20):
    plot_sample(X_train, y_train, i)
    plt.show()
```

```python
# Normalizing train and test
X_train = X_train/255
X_test= X_test/255
```

```python
#reshaping to add third dimension
X_train = X_train.reshape(-1,28,28,1)
X_train.shape
```

```python
X_test = X_test.reshape(-1,28,28,1)
X_test.shape
```

# CNN

```python
cnn = tf.keras.models.Sequential()

# adding CNN layers to the Neural Network
# CNN layer 1
cnn.add(tf.keras.layers.Conv2D(input_shape = (28, 28, 1), filters=64, kernel_size=(3,3), strides=(1,1), activation='relu'))   
cnn.add(tf.keras.layers.MaxPooling2D((2,2)))
# CNN layer 2
cnn.add(tf.keras.layers.Conv2D(filters=64, kernel_size=(3,3), strides=(1,1), activation='relu'))   
cnn.add(tf.keras.layers.MaxPooling2D((2,2)))

# adding dense layers to the Neural Network
cnn.add(tf.keras.layers.Flatten())  
cnn.add(tf.keras.layers.Dense(64, activation='relu'))  
cnn.add(tf.keras.layers.Dense(32, activation='relu'))
cnn.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy']) 
```

```python
cnn.fit(X_train, y_train, epochs=10)
```

```python
y_pred = cnn.predict(X_test)
```

```python
y_classes = [np.argmax(i) for i  in y_pred]
y_classes[0:5]
```
