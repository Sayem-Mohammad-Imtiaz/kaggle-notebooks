# 1.3) "*How much does our vote count ?*"

```python
population = pd.DataFrame({'State':['Sabah', 'Perak', 'Johor', 'Selangor', 'Kelantan', 'Pahang', 'Pulau Pinang', 'Kedah', 'Negeri Sembilan', 'Terengganu', 'Melaka', 'Sarawak', 'Perlis', 'Wilayah Persekutuan (KL)', 
                                    'Wilayah Persekutuan (Putrajaya)', 'Wilayah Persekutuan (Labuan)'], 'Seat':[85, 83, 82, 78, 59, 56, 53, 51, 44, 40, 34, 31, 18, 11, 1, 1], 
                           'Population':[3903400, 2512100, 3764300, 6528400, 1885700, 1674600, 1774600, 2180600, 1130300, 1245700, 930700, 2812800, 254400, 1780700, 103800, 99300]})
population
```

```python
seats = pd.read_csv("../input/malaysia-ge14-election-results-parliament/Election-Results-2018 - Total_Votes_by_Seat.csv")
seats['Total Registered Votes']= seats['Total Registered Votes'].str.replace(",","").astype(int)
seats['Votes for ALL Candidate']= seats['Votes for ALL Candidate'].str.replace(",","").astype(int)
seats['Spoilt Votes']= seats['Spoilt Votes'].str.replace(",","").astype(int)
seats
```

```python
# Combining the two datasets into one table for better observation 
votes = seats.groupby("State")["Votes for ALL Candidate",'Spoilt Votes',"Total Registered Votes"].sum().reset_index()
result = pd.merge(votes, population, how="inner", on=["State", "State"])
result
```

## <u>Results</u>

```python
result['Number of votes needed to win'] = (result['Total Registered Votes'] - result['Spoilt Votes']) / result['Seat']

Average = result['Number of votes needed to win'].mean()
print(f'Average number of votes required for one seat :{Average:.0f}')
```

```python
print('Number of Parliament Seat :',seats[(seats.Type == 'PARLIMEN')].count()['Seat'])
print('Number of DUN Seat :',seats[(seats.Type == 'DUN')].count()['Seat'])
```

```python
Total = Average*112
f"Number of Votes needed to form a Federal Government:  {Total:.0f}"
```

```python
data = youth.dropna()
speak = data.loc[:,['Assertiveness',"Number of friends","Public speaking","Fear of public speaking","Finances","Spending on looks","Branded clothing","Empathy","Elections","Decision making","Gender","all"]]
speak
```

### <u>Table 2.12) What are the correlations between the overall factors of youths in public speaking (using diagonal correlation matrix)</u>

```python
speak2 = speak.iloc[:,:9]
corr = speak.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5},annot=True)
```

### <u>Table 2.12) What are the correlations between the overall factors of youths in public speaking (using correlation heatmap) </u>

```python
plt.figure(figsize=(12,8))
cmap=sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(speak2.corr(), annot=True,cmap="coolwarm",vmax=.2, center=0)
```

## Observation 
>### <u>**Correlation coefficients**</u> are indicators of the strength of the linear relationship between two different variables, x and y. A linear correlation coefficient that is greater than zero indicates a <span style="color:red">positive relationship</span>. A value that is less than zero signifies a <span style="color:blue">negative relationship</span>. 

>### So by logic , Its correlation fo youths who spend on looks vs youth who spend on branded clothing has a very high positive realtionship .As well as people who are well prepared before a public speaking has a definite relationship with peeople who have a fear of public speaking.

## "*Is there a difference in Height between Youths living in Urban and Rural Areas*" 


### <u>Table 2.22) Variation of Height Difference between Village and City</u>  

```python
youth["all"] = ""
fig = plt.subplots(figsize=(12, 5))
data = youth.dropna(subset=['Height'])
sns.violinplot(x='Height', y = "all", hue='Village - town', data=data, split=True)
```

## <u>Outliers</u> 
>### As you can see theres a huge outlier where we are counting all the values of various heights but sadly its very skewed since were taking into account of extreme values . These are called outliers and as much as we want to make sure everyone is inclusive , we also want to make a good observation on what the majority of data hence its often a good idea to get rid of extreme values such as outliers.
