# Import Python Libraries

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)


import os

# charts:
import matplotlib.pyplot as plt
import seaborn as sns 
color = sns.color_palette()

# ml 
import sklearn.metrics as metrics


# h2o
import h2o
from h2o.automl import H2OAutoML
```

# Read the files in the directory

```python
data = "../input/loan-default-prediction/Default_Fin.csv"
```

# prepare data

```python

h2o.init()

h2o_train = h2o.import_file(data)
h2o_train['Defaulted?'] = h2o_train['Defaulted?'].asfactor()
```

# run model

```python

features = [x for x in h2o_train.columns if x not in ['Index', 'Defaulted?']]

h2o_aml = H2OAutoML(
    max_runtime_secs=60,
    stopping_metric='AUCPR' #AUCPR (Area Under the Precision-Recall Curve) 
)

h2o_aml.train(x=features, y='Defaulted?', training_frame=h2o_train)
```
