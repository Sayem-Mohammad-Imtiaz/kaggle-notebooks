## Analysing the data
The basic `plot()` command for a data set can be used to create a time series. In the example below the *Years* field is used for the x-axis and the different regions are plotted by listing them for the parameter y.
* Run the code in the box below to plot the time series

```python
# create a plot with Years on the x axis and the regions on the y axis 
male_le_data.plot(x='Years',y=['North East',
                               'North West',
                               'Yorkshire and The Humber', 
                               'East Midlands','West Midlands',
                               'East of England',
                               'London',
                               'South East',
                               'South West' ],
                  figsize=(10, 10))
plt.show()
```
