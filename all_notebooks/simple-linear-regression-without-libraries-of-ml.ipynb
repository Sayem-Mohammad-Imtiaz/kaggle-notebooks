```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('../input/salary/Salary.csv')
df = pd.DataFrame(df)

sns.pairplot(df)

df3 = df.copy()
```

# ***Normalizing***

```python
# NORMALIZING the data
salary = []
exp = [] 

avg_s = sum(df['Salary'])/len(df['Salary'])
range_s = max(df['Salary']) - min(df['Salary'])
for i in df['Salary']:
    salary.append((i-avg_s)/range_s)

avg_e = sum(df['YearsExperience'])/len(df['YearsExperience'])
range_e = max(df['YearsExperience']) - min(df['YearsExperience'])
for i in df['YearsExperience']:
    exp.append((i-avg_e)/range_e)

df3['Salary'] = salary
df3['YearsExperience'] = exp

# dependent and independent variables.
X = list(df3['YearsExperience'])
Y = list(df3['Salary'])
```

# **The Model**

```python
a,b = 0,0             # Weights
n = 500               # no. of iterations
n1 = len(X)           # no. of data points
r = 0.005             # learning rate
p = []

for j in range(n): 
    
    sq_error = 0          # Squared Error
    error_list = []       # Error List
    sq_error_list = []    # Squared Error List
    predicted2 = []       # Collecting Prediction to Visualize
    
    for i in range(n1) : 
        
        h = a + b*X[i]       # Hypothesis (predicted)
        e = h - Y[i]         # Error
        sq_error += e**2     # Squared Error
        
        error_list.append(e)                # Error List
        predicted2.append(h)                # Predicted List
        sq_error_list.append(e**2)          # Squared Error List   
    
    zz = [error_list[i]*X[i] for i in range(n1)]  # the product of ith term of X(feature) with it's respective Error
    
    # Weight Update
    a -= r*(sum(error_list))
    b -= r*(sum(zz))
    
    p.append(sum(sq_error_list)*(1/n1))     #collecting mean squared error to visualize the descent in error
    
```

***Weights,Prediction(last data point),Mean Squared Error,Derivative (last reached)***

```python
    print(f'Derivative Term      : {(sum(error_list))*(sum(X))*(1/n1)}')
    print(f"Mean Squared Error   : {(sum(sq_error_list)*(1/n1))}")
    print(f"Weights              : {(round(a,4), round(b,4))}")
    print(f"Predicted : {round(h,4)}, Actual : {round(Y[i],4)}")
```

# ***Visualizing***

```python
plt.xlabel('No. of Iterations')
plt.ylabel('Cost')
plt.plot([i for i in range(n)] , p)   # error with respect to iteration
plt.show()

plt.xlabel('Salary')
plt.ylabel('Experience')
plt.plot(X,Y,'.',color='red')         # the original data points
plt.plot(X,predicted2,color='blue')   # predicted line
plt.show()
```

# **Prediction on new Values**

```python
# Predictions 
input_exp = 14.5/avg_e                         # input experience (dividing by its average to work with our predicted hypothesis)
output_salary = (a + b*input_exp )*avg_s    # output salary    (multiplied by its average to restore the original value (without normalizing))
print(f"Salary Predicted : {output_salary}")
```
