H-1B Visa = It is one of the categories of short-term work visas in the US that can be applied for through sponsorship of businesses of a certain size or larger located in US territory. It is a visa that allows you to work as a foreign worker in the US. In order **to apply for this visa, you need a bachelor's degree or higher, a professional bachelor's degree and at least 6 years of work experience.** The major and job of the degree must match, and must be a profession other than a nurse. Nurses and primary industry workers are classified under a visa category other than H-1B.

# Pre-Processing

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        
import matplotlib.pyplot as plt   
from matplotlib.pyplot import pie
import seaborn as sns

#Set Matplotlib defaults
plt.rc('figure', autolayout=True)
plt.rc('axes', labelweight='bold', labelsize='large')
plt.rc('image', cmap='magma')

import os, warnings
warnings.filterwarnings("ignore") # to clean up output cells
```

```python
CSV = '/kaggle/input/h1b-visa-petitions-20152019/h1b_disclosure_data_2015_2019.csv'

df = pd.read_csv(CSV)

conditions = [
    (df['WORKSITE_STATE_FULL'] == 'INDIANA'),
    (df['WORKSITE_STATE_FULL'] == 'WISCONSIN'),
    (df['WORKSITE_STATE_FULL'] == 'MINNESOTA'),
    (df['WORKSITE_STATE_FULL'] == 'OHIO'),
    (df['WORKSITE_STATE_FULL'] == 'MICHIGAN'),
    (df['WORKSITE_STATE_FULL'] == 'PENNSYLVANIA'),
    (df['WORKSITE_STATE_FULL'] == 'ILLINOIS'),
    (df['WORKSITE_STATE_FULL'] == 'NEW YORK') ]
choices = [1,2,3,4,5,6,7,8]
df['sorter'] = np.select(conditions, choices, default=0)

df = df.sort_values(by='sorter')
```

The data preprocessing part is nothing special to comment on.

# EDA

```python
fig = plt.figure(figsize=(7,7))

fig.patch.set_facecolor('#F1FBFF')
fig.patch.set_alpha(1.0)

sums = df.CASE_NUMBER.groupby(df.CASE_STATUS).count()
pie(sums, labels=sums.index, autopct='%.1f%%', textprops={'fontsize': 12})

ax1 = plt.title('Case Status of H-1B Visa',
          fontsize = 16,
          fontweight = 'heavy',
          loc='center', 
          pad=30); #semi-colon for hide text before graph output
```

On the dataset, most of cases are certified (89%)

```python

fig = plt.figure(figsize=(12,12))

fig.patch.set_facecolor('#F1FBFF')
fig.patch.set_alpha(1.0)

fig.tight_layout() 
fig.subplots_adjust(hspace=.4)

ax1 = plt.subplot(2,2,1)

ax1 = sns.countplot(x="YEAR", data=df[df['CASE_STATUS']=="CERTIFIED"], palette= "Blues")
ax1.patch.set_facecolor('#F1FBFF')

ax1 = plt.gca().spines['right'].set_visible(False)
ax1 = plt.gca().spines['top'].set_visible(False)
ax1 = plt.gca().spines['left'].set_visible(True)
ax1 = plt.gca().spines['bottom'].set_visible(True)

plt.title('Unit: #case\n2015-2019', fontsize = 9, fontweight = 'light',loc='right')
plt.title('Cetified Cases of H-1B Visa',
          fontsize = 13,
          fontweight = 'heavy',
          loc='center', 
          pad=30); #semi-colon for hide text before graph output

ax2 = plt.subplot(2,2,2)

ax2 = sns.countplot(x="YEAR", data=df[df['CASE_STATUS']=="CERTIFIED-WITHDRAWN"], palette= "Blues")
ax2.patch.set_facecolor('#F1FBFF')

ax2 = plt.gca().spines['right'].set_visible(False)
ax2 = plt.gca().spines['top'].set_visible(False)
ax2 = plt.gca().spines['left'].set_visible(True)
ax2 = plt.gca().spines['bottom'].set_visible(True)

plt.title('Unit: #case\n2015-2019', fontsize = 9, fontweight = 'light',loc='right')
plt.title('Cetified-Withdrawn of H-1B Visa',
          fontsize = 13,
          fontweight = 'heavy',
          loc='center', 
          pad=30); #semi-colon for hide text before graph output

ax3 = plt.subplot(2,2,3)

ax3 = sns.countplot(x="YEAR", data=df[df['CASE_STATUS']=="WITHDRAWN"], palette= "Blues")
ax3.patch.set_facecolor('#F1FBFF')

ax3 = plt.gca().spines['right'].set_visible(False)
ax3 = plt.gca().spines['top'].set_visible(False)
ax3 = plt.gca().spines['left'].set_visible(True)
ax3 = plt.gca().spines['bottom'].set_visible(True)

plt.title('Unit: #case\n2015-2019', fontsize = 9, fontweight = 'light',loc='right')
plt.title('Withdrawn Cases of H-1B Visa',
          fontsize = 13,
          fontweight = 'heavy',
          loc='center', 
          pad=30); #semi-colon for hide text before graph output

ax4 = plt.subplot(2,2,4)

ax4 = sns.countplot(x="YEAR", data=df[df['CASE_STATUS']=="DENIED"], palette= "Blues")
ax4.patch.set_facecolor('#F1FBFF')

ax4 = plt.gca().spines['right'].set_visible(False)
ax4 = plt.gca().spines['top'].set_visible(False)
ax4 = plt.gca().spines['left'].set_visible(True)
ax4 = plt.gca().spines['bottom'].set_visible(True)

plt.title('Unit: #case\n2015-2019', fontsize = 9, fontweight = 'light',loc='right')
plt.title('Denied Cases of H-1B Visa',
          fontsize = 13,
          fontweight = 'heavy',
          loc='center', 
          pad=30); #semi-colon for hide text before graph output

```

Cetified cases keep its trend.

WITHDRAWN related cases decreased on 2019.

Denied cases had keep decrased from 2015.
