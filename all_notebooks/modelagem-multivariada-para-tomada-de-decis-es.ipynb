```python


#sort to bring highest grosser movie to the top bring movie title column to front
imdb_dataByGross = imdb_data.sort_values(by = 'gross', ascending = False) 

imdb_dataByGross.head()
```

```python
#cleaning the missing data inserting 0
imdb_gross_missing = imdb_data['gross'].fillna(0)

#after clean
imdb_gross_clean = imdb_data['gross'].count()

print ('Gross without Nan = ',imdb_gross_clean)
print ('Gross with Nan = ',imdb_gross_missing.count())
print ('Missing Data Gross = ', imdb_gross_missing.count() - imdb_gross_clean)
```

```python
# population_review % critic_review. Does it affect with incomming gross?
tst = (imdb_data['movie_title'],imdb_data['movie_facebook_likes']/imdb_data['imdb_score'])

movie_title = tst[0]
score = tst[1]
d= {'movie_title': imdb_data['movie_title'],
    'likes':imdb_data['movie_facebook_likes'],
    'gross': imdb_data['gross'],
    'duration': imdb_data['duration'],
    'budget': imdb_data['budget'],
    'lucro': imdb_data['gross'] - imdb_data['budget'],
    'director': imdb_data['director_name']
   }
df = pd.DataFrame(d)
#print (dataframe.sort_values(['likes'], ascending = [False]))
df.plot(x=['gross'],y=['likes'], kind= 'scatter' )
```
