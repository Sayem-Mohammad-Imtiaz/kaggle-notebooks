#country, iso_code, date, vaccine, source_name & source_website type should change as they are non-numeric

```python
cv_data['country'] = cv_data['country'].astype(str)
cv_data['iso_code'] = cv_data['iso_code'].astype(str)
cv_data['date'] = pd.to_datetime(cv_data['date'], format="%Y/%m/%d")
cv_data['vaccines'] = cv_data['vaccines'].astype(str)
cv_data['source_name'] = cv_data['source_name'].astype(str)
cv_data['source_website'] = cv_data['source_website'].astype(str)
cv_data.head()
```

```python
cv_data[cv_data['iso_code'] == 'CAN'].head()
```

```python
cols = ['total_vaccinations','people_vaccinated','people_fully_vaccinated', 'daily_vaccinations_raw', 'daily_vaccinations', 'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred', 'daily_vaccinations_per_million']
df = cv_data.copy()
df.update(df.groupby('country')[cols].ffill().fillna(0))
df.head()
```
