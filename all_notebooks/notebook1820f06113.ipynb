```python


filepath = "/kaggle/input/rigadsclub/linkedin/riga-ds-club_updates_engagement_metrics.csv"
engagement = pd.read_csv(filepath, parse_dates=True)
engagement.head(1)

```

```python
dayofweek = pd.to_datetime(engagement['Created date'], infer_datetime_format=True)
#print(dayofweek.dt.dayofweek)

engagement['dayofweek'] = dayofweek.dt.dayofweek.astype('int')
engagement['month'] = dayofweek.dt.month.astype('int')

forHeatmap = engagement.pivot_table(index='dayofweek', columns='month', values='Likes')


ax = sns.heatmap(forHeatmap, cmap="YlGnBu")
```
