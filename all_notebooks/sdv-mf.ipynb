```python
a= array([[4,0],[3,-5]])
m=2
n=2
print(a)
u,s,v=svd(a)

print(u)
print(s)
print(v)
```

```python
at=a.T
A=at.dot(a)
#print(A)

w,v =np.linalg.eig(A)
#print(w)
#print(v)

s = np.sqrt(w)
S=np.zeros((n,n))
for i in range(n):
    S[i][i]=s[i]

Sinv = np.linalg.inv(S)
print(S)

Vt =v.T
print(Vt)


U=a.dot((v.dot(Sinv)))
print(U)
```

```python


Ans= U.dot((S.dot(Vt)))
print(Ans)
```
