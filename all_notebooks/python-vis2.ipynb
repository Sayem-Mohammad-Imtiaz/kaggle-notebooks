```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('seaborn-deep')

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("../input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv")
data.head(5)
```

# Data Cleaning

# Data Visualization

```python
plt.rcParams['figure.figsize'] = (10, 5)
sns.countplot(data['sex'], palette = 'pink')
plt.title('Sex of the Patients', fontsize = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
sns.countplot(data['smoking'], palette = 'pink')
plt.title('Smoking Conditions of the Patients', fontsize = 20)
plt.show()
```

```python
labels = ['0', '1']
sizes = data['high_blood_pressure'].value_counts()
colors = plt.cm.copper(np.linspace(1, 0, 2))

plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(sizes, labels = labels, colors = colors, shadow = True)
plt.title('High Blood Pressure of Patients', fontsize = 20)
plt.legend()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
sns.countplot(data['age'], palette = 'pink')
plt.title('Ages of the Patients', fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10, 5)
sns.countplot(data['DEATH_EVENT'], palette = 'pink')
plt.title('Death probability of the Patients', fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (30, 20)
sns.heatmap(data.corr(), annot = True)

plt.title('Histogram of the Dataset', fontsize = 30)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 7)
ax = sns.violinplot(x = data['sex'], y = data['smoking'], palette = 'Reds')
ax.set_xlabel(xlabel = 'Sex', fontsize = 9)
ax.set_ylabel(ylabel = 'Smoking', fontsize = 9)
ax.set_title(label = 'Smoking Conditions of patients from different sexes', fontsize = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 7)
ax = sns.violinplot(x = data['ejection_fraction'], y = data['DEATH_EVENT'], palette = 'Reds')
ax.set_xlabel(xlabel = 'ejection fraction', fontsize = 9)
ax.set_ylabel(ylabel = 'DEATH_EVENT of Patients', fontsize = 9)
ax.set_title(label = 'DEATH_EVENT of Patients with different ejection fraction', fontsize = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 7)
ax = sns.violinplot(x = data['serum_creatinine'], y = data['DEATH_EVENT'], palette = 'Reds')
ax.set_xlabel(xlabel = 'serum_creatinine ', fontsize = 9)
ax.set_ylabel(ylabel = 'DEATH_EVENT of Patients', fontsize = 9)
ax.set_title(label = 'DEATH_EVENT of Patients with different serum_creatinine ', fontsize = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 7)
ax = sns.violinplot(x = data['serum_sodium'], y = data['DEATH_EVENT'], palette = 'Reds')
ax.set_xlabel(xlabel = 'serum_sodium ', fontsize = 9)
ax.set_ylabel(ylabel = 'DEATH_EVENT of Patients', fontsize = 9)
ax.set_title(label = 'DEATH_EVENT of Patients with different serum_sodium ', fontsize = 20)
plt.show()
```
