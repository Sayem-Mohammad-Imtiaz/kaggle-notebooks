# Fantastic Pokemons and where to find them

A small data analysis to discover, from the rarest and cutest Pokemon to the almighty ones, all the curiosities and statistics you did not know before.

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
%matplotlib inline
import matplotlib.pyplot as plt

plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (13.0, 6.5)

# Any results you write to the current directory are saved as output.
df = pd.read_csv('../input/Pokemon.csv')
```

```python
df.head()
```

```python
print('We have overall', len(df.values), 'Pokemons, from a total of', df['Generation'].max(), 'generations.')
```

```python
pokemon_types = df.groupby(['Type 1'])['Name'].count()

print('Different types:', pokemon_types.count())

fig = plt.figure()
fig.suptitle('Pokemon types')
plt.subplot(1, 2, 1)
pokemon_types.plot(kind='bar')
plt.subplot(1, 2, 2)
pokemon_types.plot(kind='pie')
plt.show()
```

```python
pokemon_legendary_types = df[df['Legendary'] == True].groupby(['Type 1'])['Name'].count()

print('Different types:', pokemon_legendary_types.count())

fig = plt.figure()
fig.suptitle('Pokemon types')
plt.subplot(1, 2, 1)
pokemon_legendary_types.plot(kind='bar')
plt.subplot(1, 2, 2)
pokemon_legendary_types.plot(kind='pie')
plt.show()
```

```python
fig = plt.figure()
fig.suptitle('Percentage of legendary Pokemon per type')
(pokemon_legendary_types / pokemon_types).plot(kind='bar')
plt.show()

fig = plt.figure()
fig.suptitle('Pokemon types')
pokemon_types.plot(kind='barh', color='green')
pokemon_legendary_types.plot(kind='barh')
plt.show()
```
