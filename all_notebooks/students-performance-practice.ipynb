```python
#What group obtained the best average score?
#Group by group, the obtain the average of the columns separated by group. Average again the three columns. race/ethnicity

attributesMean  = ['math score', 'reading score', 'writing score']
studentsDfGrouped = studentsDf.groupby(['race/ethnicity']).mean(attributesMean)
studentsDfGrouped['Overall Mean Score'] = studentsDfGrouped.loc[:, 'math score': 'writing score'].mean(axis=1)
print(studentsDfGrouped)

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns


plt.figure(figsize = (10,6))
plt.title('Overall Mean Score per Group')
sns.barplot(x=studentsDfGrouped.index, y=studentsDfGrouped['Overall Mean Score'])

plt.xlabel('Groups')
plt.ylabel('Overall Mean Score')
```

```python

#What group obtained the best math average score?
plt.figure(figsize = (10,4))
plt.title('Group with the best math average score')
a = sns.barplot(x=studentsDfGrouped.index, y=studentsDfGrouped['math score'], palette="Blues_d")
plt.xlabel('Groups')
plt.ylabel('Math Average Score')


##What group obtained the best reading score?
plt.figure(figsize = (10,4))
plt.title('Group with the best reading average score')
b = sns.barplot(x=studentsDfGrouped.index, y=studentsDfGrouped['reading score'], palette="dark:salmon")
plt.xlabel('Groups')
plt.ylabel('Reading Average Score')


#What group obtained the best writing score?
plt.figure(figsize = (10,4))
plt.title('Group with the best writing average score')
sns.barplot(x=studentsDfGrouped.index, y=studentsDfGrouped['writing score'], palette="Blues_d")
plt.xlabel('Groups')
plt.ylabel('Writing Average Score')
```
