```python
#Reading data from the input file
data=pd.read_csv('../input/inc_occ_gender.csv')
```

```python
#Filtering the list of industries from the dataset
all_industries=[industry for industry in data.Occupation if industry.isupper()]
```

```python
all_industries
```

```python
#Getting the industries data alone into a separate dataframe
all_industries_data=data.loc[data['Occupation'].isin(all_industries)]
```
