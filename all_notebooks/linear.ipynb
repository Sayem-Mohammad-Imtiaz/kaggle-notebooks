```python
def outlier_treatment(Area):
    sorted(Area)
    Q1,Q3 = np.percentile(Area , [25,75])
    IQR = Q3 - Q1
    lower_range = Q1 -(1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range,upper_range
lowerbound,upperbound = outlier_treatment(data.Area )
```

```python
data[(data.Area < lowerbound) | (data.Area > upperbound)]
```

```python
data.drop(data[(data.Area < lowerbound) | (data.Area > upperbound)].index , inplace=True)
```

```python
normalized = preprocessing.normalize(data)
n_data=pd.DataFrame(normalized,columns=['Area','BHK','Bathroom','Furnishing','Parking','Price','Status','Transaction','Type','Per_Sqft'])
```

```python
data1_x=data1.copy()
data1.head()
```

```python
data1_x.drop(['Price'],axis=1,inplace=True)
```

```python
data1_y=data1['Price'].copy()
```

```python
data1_y.head()
```

```python
train_data,test_data,train_Target,test_Target = model_selection.train_test_split(data1_x, data1_y, random_state = 10)
```

```python
model = LinearRegression().fit(train_data, train_Target)
```

```python
r_sq = model.score(test_data, test_Target)
print('coefficient of determination:',r_sq)
```

```python
print('intercept:', model.intercept_)
print('slope:', model.coef_)
```
