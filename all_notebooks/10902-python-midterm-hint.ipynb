大樂透是一種樂透型遊戲。您必須從01~49中任選6個號碼進行投注。開獎時，開獎單位將隨機開出六個號碼加一個特別號，這一組號碼就是該期大樂透的中獎號碼，也稱為「獎號」。您的六個選號中，如果有三個以上（含三個號碼）對中當期開出之六個號碼（特別號只適用於貳獎、肆獎、陸獎和柒獎），即為中獎，並可依規定兌領獎金。

````python
#第一題程式由下方 pass 處開始寫 -- 刪除 pass 
#中獎號 win_no 特別號 s_no 題目已給定 
def lottery_check(numbers, special):
    '''
    中獎號 [1,2,3,4,5,6] 特別號 7
    >>> lottery_check([1,2,11,22,44,12], 7)
    >>> 柒獎
    >>> lottery_check([1,2,3,4,5,6], 12)
    >>> 頭獎
    '''
    # ==========================================
    #  SOLUTION SAMPLE START
    # ==========================================
    ans = rewards[8] # 一開始沒中
    right_no = False # right_no 一開始設成不合法選號
    
    if len(numbers) <= 6 : # 如果 nubers list 的大小 <= 6 
        right_no = True # right_no 設成合法選號
    win_cnt = 0 # 一開始沒有數字中獎
    for no in numbers:   # 對於每一個在選號 numbers 中的數字 no
        if no in win_no : # 如果選號的數字 no 在獎號 win_no 中
            win_cnt += 1 # 對中獎號的數目 +1
            
    got_special = False  # 一開始沒中特別號
    if special == s_no: # 
        got_special = True # 對中特別號 
    
    if right_no == True : # 合法選號
        if win_cnt == 4 and got_special == True: # 4 + 1
            ans = rewards[3] # 肆獎
        elif win_cnt == 4:  # 4
            ans = rewards[4] # 伍獎
        elif win_cnt == 3: # 3
            ans = rewards[6] # 普獎
        else: # 沒中 => 中獎條件之外的情形
            ans = rewards[8] # 沒中
    # ==========================================
    #  SOLUTION SAMPLE END 
    # ==========================================
    return ans

#第一題程式測試 --- 下列程式請勿修改
#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!
def GetChkNo(ns): # generate ns(nuber size) check numbers
    nums = win_no[0:ns]  # get win_no slice
    while len(nums) != 6: # add numbers to nums till the nums length is 6  
        no = random.randint(1,49)
        if no != s_no and no not in win_no and no not in nums:
            nums.append(no)
    random.shuffle(nums)
    return nums
#check problem1 
print(f"中獎 號碼： {sorted(win_no)} 特別號 {s_no}")
print("========================================================")
for i in range(9):
    if i in rws:     # the reward in rws is need to be checked
        slice_no = (i+1) // 2  # (i+1)//2 is the size of the numbers are need to be detele 
                               # reward 0->0; 1,2->1; 2,3->2 ... 
        win_size = len(win_no)-slice_no # the win_no size need to reserve 
        sno = s_no
        if i%2 == 0:  # the rewards don't have special number set it to 0
            sno = 0
        nums = GetChkNo(win_size)
        print(f"{rewards[i]} 號碼： {sorted(nums)} 特別號：{sno} 結果： {lottery_check(nums, sno)}")
#!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!  
````
