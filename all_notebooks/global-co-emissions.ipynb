```python
#import functional packages and libraries

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns # data visualisation library
import matplotlib.pyplot as plt #matlab plotting library
import plotly.express as px #plotly's smart graphs

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#read csv file (kaggle data set) on pd dataframe
filename = "/kaggle/input/co2-ghg-emissionsdata/co2_emission.csv"
df = pd.read_csv(filename)
df.head() #show first 5 rows of dataframe
```

```python
df.describe() #quick statistics with describe function
```

```python
#plot emissions for all countries, continents and entities with the color shade indicating the year of recorded emissions.
fig = px.scatter(df, x='Annual CO₂ emissions (tonnes )', y='Entity',color='Year')
fig.show()
```

```python
#filtering the continents, transport and stat. differences from the full dataframe
world_continents = df['Entity'].isin(['World','Africa','EU-28','Americas (other)','Asia and Pacific (other)',
                                      'Middle East','Statistical differences','International transport'])

#plot all the countries without the continents and entities (note the negative emissions in the graph 
#above have disappeared in the one below)
fig = px.scatter(df[~world_continents], x='Annual CO₂ emissions (tonnes )', y='Entity',color='Year')
fig.show()
```

```python
#plot only the continents, transport, stat. differences and world values
fig = px.scatter(df.loc[world_continents], x='Entity', y='Annual CO₂ emissions (tonnes )',color='Year')
fig.show()
```

Data is suspicious as the World values do not match the sum of the reported continents, 
international transport and statistical differences. 

Moreover, just the USA in 2017 emitted 5.26 B tonnes of CO2, while the reported value for Americas (other) 
continent in 2014 is just 2.4 B tonnes of CO2.

```python
def emissions_by_entity_in_year(entity,year):
    
    return print(df.loc[df['Entity'].isin([entity]) & 
                        
                        df['Year'].isin([year])]['Annual CO₂ emissions (tonnes )'])
```
