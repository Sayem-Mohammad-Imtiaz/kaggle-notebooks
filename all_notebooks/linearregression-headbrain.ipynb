Linear Regression Model_1: HeadBrain dataset

```python
import pandas as pd
import numpy as np
```

libraries imported

```python
df=pd.read_csv("../input/headbrain-lr/headbrain.csv")
```

reading dataset through pandas and storing dataset in 'df' variable

```python
df.shape
```

```python
from sklearn.linear_model import LinearRegression
```

To import LinearRegression class from sklearn library

```python
lr_model=LinearRegression()
```

```python
x=df.iloc[:,0:1].values #range to convert data from one d array to 2d array
y=df.iloc[:,1].values
lr_model.fit(x,y)
```

x should be in 2D array as independent variable is of 2d array.
acc. to: y=mx+c
y is of 1D array

```python
#Since y=mx+c:
m=lr_model.coef_  #slope
c=lr_model.intercept_ #intercept
```

```python
m,c
```

```python
y_predict=(m*x)+c
```

```python
y_predict #predicted value of y
```

.

**Now, plotting the graph for predicted value vs actual value**

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y) #blue dot=actual value
plt.plot(x,y_predict,c="red") #red line=predicted value
plt.show()
```

```python
y_predict=lr_model.predict(x) #y_predict=mx+c
```

```python
#suppose
head=4002
bweight=lr_model.predict([[head]])
```

```python
plt.scatter(x,y)  #actual data with blue dots
plt.plot(x,y_predict,c="red") #prediction line
plt.scatter([head],bweight,c="yellow") #predicted value for input data
plt.show()
```

Conclusion: 
The yellow dot on red line shows that our data is trained and can be used to predict the head size and brain weight for any given value.
