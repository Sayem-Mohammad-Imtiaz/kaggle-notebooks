```python
main_big_five_data = big_five_data.copy()
main_big_five_data = main_big_five_data.iloc[:, 51:]
main_big_five_data.loc[:, 'ACRS'] = ''
name_columns = ['EXT', 'EST', 'AGR', 'CSN', 'OPN']
var_degree = ['1', '2', '3']
main_big_five_data.head()
```

```python
main_big_five_numpy = main_big_five_data.to_numpy()
```

```python
import time
start = time.time()
size_frame = len(main_big_five_numpy) 
i = 0
mas = []
for row in main_big_five_numpy:
    calc_acrs = ''
    for j in range(5): 
        if row[j] <= lower_value:
            calc_acrs += name_columns[j] + var_degree[0]
        elif row[j] > lower_value and row[j] <= medium_value:
            calc_acrs += name_columns[j] + var_degree[1]
        else:
            calc_acrs += name_columns[j] + var_degree[2]        
    mas.append(calc_acrs)
    if ((i+1) % 1000 == 0):
        end = time.time()
        print("{}%, elapsed time: {}".format((i+1)*100.0 / size_frame, end - start), end='\r')
    i += 1
print("Complete for time: {}".format(end - start), end='\r')
```

```python
main_big_five_data.iloc[:, 5] = mas
main_big_five_data.tail()
```

```python
dict_arcs = dict()
for elem in mas:
    if elem in dict_arcs.keys():
        dict_arcs[elem] += 1
    else:
        dict_arcs[elem] = 1
        
dict_arcs = dict(sorted(dict_arcs.items(), key=lambda item: item[1], reverse=True))

dict_arcs_10 = dict()
i = 0
for elem in dict_arcs.items():
    if i >= 10:
        break
    i += 1
    dict_arcs_10[elem[0]] = elem[1]
    
dict_arcs_10
```

```python
plt.figure(figsize=(10,6))
plt.title("10 most polular ACRS clasters")
sns_diag = sns.barplot(x=list(dict_arcs_10.keys()), y=list(dict_arcs_10.values()))
```

```python
keys_10 = list(dict_arcs_10.keys())
values_10 = list(dict_arcs_10.values())
print('Name claster: {} from {}, percent: {}%'.format(keys_10[0], values_10[0], 100.0 * values_10[0]/len(main_big_five_data)))
```

```python
name_columns = ['EXT', 'EST', 'AGR', 'CSN', 'OPN']
var_degree = ['1', '2', '3']

def signal_from_mas(num):    
    state = ''
    md = num
    for i in range(len(name_columns)):        
        state += name_columns[i] + var_degree[md % 3]
        md = md // 3
    return state
```

```python
clusters = dict()
for i in range(243):
    state = signal_from_mas(i)
    if state in dict_arcs:
        clusters[i] = dict_arcs[state]
    else:
        clusters[i] = 0
```

```python
plt.figure(figsize=(18,6))
plt.title("Clusters and The quantity of people in it")
sns_diag = sns.barplot(y=list(clusters.values()), x=list(clusters.keys()))
```
