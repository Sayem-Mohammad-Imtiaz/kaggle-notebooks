```python
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#loading data to cook!
stud_filepath = "../input/students-performance-in-exams/StudentsPerformance.csv"
stud_data = pd.read_csv(stud_filepath)
stud_data #reviewing data to work with 
```

So I wanted to find out whether kind of lunch affect scores or not.
I created a new column and named it "total scores". As the name suggests, it is the total of all three scores(math,reading and writing). 

```python
stud_data['total score'] = stud_data['math score'] + stud_data['reading score']+ stud_data['writing score']
sns.swarmplot(x=stud_data['lunch'],y=stud_data['total score'])
```

As you can see from the fig, (people who get)standard lunch got better(slightly) scores than free/reduced!
But I still wanted to visualise properly. So I just took a look at more ways by which I can plot categorical data(here's the link https://seaborn.pydata.org/tutorial/categorical.html).
I did found one and also compared race/ethnicity. And again, standard lunch did better.



```python
sns.catplot(x="lunch",y="total score",hue="race/ethnicity",kind="bar",data=stud_data)
```
