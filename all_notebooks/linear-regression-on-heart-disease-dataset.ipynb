```python
# LINEAR REGRESSION ON 'heart_disease' DATASET TO COMPARE ORIGINAL RESULT AND PREDICTED RESULT
# WE WILL USE 'age' TO 'thal' COLUMNS FOR ACCURACY PREDICTION
# AS YOU CAN OBSERVE THAT ACCURACY IS DECREASING WHEN WE REMOVE DEPENDENT COLUMNS

import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn import model_selection

df=pd.read_csv('../input/heart-disease-patients-details/heart_disease.csv')

Y=df['result']
X=df.loc[:,'age':'thal']

X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,test_size=0.33,random_state=4)

lr=LinearRegression()
lr.fit(X_train,Y_train)
abc=lr.predict(X_test)

print('The Accuracy Score is {}'.format(lr.score(X_train,Y_train)))
print(abc)
print(Y_test)
```
