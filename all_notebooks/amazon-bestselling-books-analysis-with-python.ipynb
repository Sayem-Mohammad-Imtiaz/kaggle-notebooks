```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv('../input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
df.sample(5)
```

```python
# Top rated books (Revieved by at least 3000 people)
data_order = df.groupby(['Name', 'Author', 'Genre'], as_index=False)[['User Rating', 'Reviews']].mean()
data_order = data_order[data_order['Reviews']>3000]
data_order = data_order.sort_values('User Rating', ascending=False)
data_order.sample(20)
```

```python
m = df['Reviews'].min()
C = df['User Rating'].mean()
def weighted_Rating(x , m=m , C=C):
    v = x['Reviews']
    R = x['User Rating']
    return (v / (v + m) * R) + (m / (v+m) * C)
```

```python
df['Weighted Rating'] = df.apply(weighted_Rating , axis = 1)
```

```python
df.groupby(['Name' , 'Genre' , 'Author'] , as_index = False)[['Reviews' , 'User Rating' , 'Weighted Rating']].mean().sort_values(by = 'Weighted Rating' , ascending = False).head(10)
```
