```python
# Continue

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# data vizualization, linear regression and logistic regresson
```

```python
data = pd.read_csv ('../input/students-performance-in-exams/StudentsPerformance.csv')
data.head()
```

```python
data = data.rename( columns = {'lunch': 'food','test preparation course':'prep', 'race/ethnicity':'race', 'gender':'gender', 'parental level of education':'par_ed'})
data.head()
```

```python
data = pd.get_dummies(data, columns = ['food','prep', 'race', 'gender', 'par_ed'])
data['score avrg'] = (data['math score'] + data['reading score'] + data['writing score']) / 3
data = data.drop(['math score', 'reading score', 'writing score'], axis = 1)
data.head()
```

```python
from sklearn.model_selection import train_test_split
x = data.drop(['score avrg'], axis = 1)
y = data['score avrg']
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size =0.2, shuffle=True)
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg = linear_reg.fit(x_train, y_train)
```

```python
linear_reg.predict(x_test)
```

```python
linear_reg.score(x_test, y_test)
```

```python
linear_reg.coef_
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

log_reg = LogisticRegression()

log_reg = log_reg.fit(x_train, y_train)
```

```python
# getting rid of the warning:

from sklearn.pipeline import make_pipeline

pipe = make_pipeline(StandardScaler(), LogisticRegression())
pipe.fit(x_train, y_train)
```
