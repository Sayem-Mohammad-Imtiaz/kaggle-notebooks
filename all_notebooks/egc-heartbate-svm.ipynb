```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv('../input/heartbeat/mitbih_train.csv', header=None)
data.head()
```

```python
data[187].value_counts()
```

```python
c0 = data[data[187] == 0].sample(1000)
c1 = data[data[187] == 1].sample(1000)
c2 = data[data[187] == 2].sample(1000)
c3 = data[data[187] == 3]
c4 = data[data[187] == 4].sample(1000)
```

```python
new_data = pd.concat([c0, c1, c2, c3, c4])
```

```python
new_data[187].value_counts()
new_data.shape
new_data = np.array(new_data)
```

```python
X = new_data[:, : 187]
Y = new_data[:, 187]
```
