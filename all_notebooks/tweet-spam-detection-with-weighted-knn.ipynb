**Import dataset**

```python
import pandas as pd
import numpy as np

tweet_df = pd.read_csv('../input/tweet-spam/data_train.csv')
tweet_df
```

Apa aja kolomnya?

```python
tweet_df.columns
```

Drop atau hapus kolom yang tidak terpakai

```python
tweet_df.drop(['Id','following','followers','actions','is_retweet','location'], axis=1, inplace=True)
```

Ubah tipe dari kolom 'Type' sehingga Quality menjadi 0 dan Spam menjadi 1

```python
tweet_df['Type']=tweet_df['Type'].map({'Quality':0, 'Spam':1})
```

Ubah Tweet dari text menjadi angka untuk memudahkan sistem dalam proses pengklasifikasian dengan memanfaatkan vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x = tweet_df['Tweet']
y = tweet_df['Type']

x = cv.fit_transform(x)
x
```

Buat data train dan data test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
x_train.shape
```

Membuat model algoritma Weighted KNN

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3, weights='distance')
model.fit(x_train, y_train)
```

Tes akurasi model dengan menggunakan data test

```python
model.score(x_test, y_test)
```
