```python
import pandas as pd
import numpy as np
# from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import pandas_profiling as pp
from sklearn.cluster import KMeans
%matplotlib inline

import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# from sklearn.tree import export_graphviz
from sklearn import decomposition
```

```python
data=pd.read_csv("/kaggle/input/heart-attack-prediction/data.csv")
data.info()
```

```python
#Fetch features of type Object
objFeatures = data.select_dtypes(include="object").columns

#Iterate a loop for features of type object
from sklearn import preprocessing
le = preprocessing.LabelEncoder()

for feat in objFeatures:
    data[feat] = le.fit_transform(data[feat].astype(str))
    
data.info()
```

```python
del data ['oldpeak']
```

```python
data.info()
```

```python
ints=[]
for i in range(1,10):
    kmeans=KMeans(n_clusters=i,init="k-means++")
    kmeans.fit(data)
    ints.append(kmeans.inertia_)

plt.figure(figsize=(8,6))
plt.plot(range(1,10),ints, marker="*")
plt.title("Elbow Method")
plt.show()
```

```python
kmeans=KMeans(n_clusters=2, init="k-means++")
X_centered = data - data.mean(axis=0)
cl=kmeans.fit_predict(X_centered)
```

```python
print(kmeans.labels_)
```

```python
kmeans.cluster_centers_
```

```python
data.groupby(cl).size() 
```

```python
pca = decomposition.PCA(n_components=3)
X_centered = data - data.mean(axis=0)
pca.fit(X_centered)
X_pca = pca.transform(X_centered)
plt.figure(figsize=(8,6))
plt.plot(X_pca[cl == 0, 0], X_pca[cl == 0, 1],'bo', label='cluster1')
plt.plot(X_pca[cl == 1, 0], X_pca[cl == 1, 1],'ro', label='cluster2')
plt.legend(loc=0);
```

```python
print('Accuracy: ',silhouette_score(data, kmeans.labels_));
```
