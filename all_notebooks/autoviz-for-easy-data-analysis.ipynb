# AutoViz for Easy EDA

AutoViz is a simple but powerful library which can do the EDA of your data much fast. It is very helpful for rapid visualisations and helps you to figure out which variables are significant. Let us use it to see the data

```python
import pandas as pd
data = pd.read_csv('../input/data-analyst-jobs/DataAnalyst.csv', index_col = 0)
data.head()
```
