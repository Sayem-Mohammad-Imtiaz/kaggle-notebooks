```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../input/moscow-schools-winners-of-educational-olympiads/moscow schools - winners.csv", encoding="UTF-8")
df.shape
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df_clean = df.fillna('No stages')
df_clean.isnull().sum()
```

```python
top10 = df_clean[["ShortName", "OlympiadType", "Status"]][(df_clean.OlympiadType == "Всероссийская олимпиада")  & (df_clean.Status == "победитель")]['ShortName']

top10 = top10.value_counts().head(10).to_frame().reset_index().rename(columns={"ShortName": "count", "index": "ShortName"})

top10_plot = sns.barplot(data=top10, x="count", y="ShortName")
top10_plot.set_ylabel("Name")
top10_plot.set_xlabel("Number of wins")
```

```python
df_clean = df_clean.replace({"НИУ ВШЭ. лицей": "лицей НИУ ВШЭ", 
                             'ГБОУ «Лицей «Вторая школа»': 'ГБОУ лицей «Вторая школа»',
                             'ГБОУ школа № 1329': 'ГБОУ лицей № 1329', 
                             'ГБОУ школа № 1535': 'ГБОУ лицей № 1535',
                             'ГБОУ школа № 179': 'ГАОУ ВО МИОО. СОШ № 179'})
```

```python
top10 = df_clean[["ShortName", "OlympiadType", "Status"]][(df_clean.OlympiadType == "Всероссийская олимпиада") & (df_clean.Status == "победитель")]['ShortName']

top10 = top10.value_counts().head(10).to_frame().reset_index().rename(columns={"ShortName": "count", "index": "ShortName"})

top10_plot = sns.barplot(data=top10, x="count", y="ShortName")
top10_plot.set_ylabel("Name")
top10_plot.set_xlabel("Number of wins")
```
