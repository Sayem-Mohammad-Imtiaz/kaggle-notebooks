# Summary of Data

```python
df = pd.read_csv('../input/students-performance-in-exams/StudentsPerformance.csv')
df.head()
```

## Column Types in the Dataset 

```python
df.info()
```

## Analysis of Marks in each Subject based on Gender
It is observed that for Female category has a better performance in Math Score compared to Male Category. However, the average scores in the other 2 subjects seemed to be higher for the Male category.

```python
df1 = df.groupby('gender')[['math score','reading score','writing score']].mean()
df1.reset_index(inplace=True)
figure1 = px.bar(df1,x='gender',y=['math score','reading score','writing score'],width=800)
figure1.show()
```

# Analysing the marks of each Subject Based on Race/Ethinicity

```python
df2 = df.groupby('race/ethnicity')[['math score','reading score','writing score']].mean()
df2.reset_index(inplace=True)
df2
figure2 = px.bar(df2,x='race/ethnicity',y=['math score','reading score','writing score'],width=800)
figure2.show()
```
