```python
plt.rcParams['figure.figsize']=(15,6)
sns.countplot(x=df['Platform'])
plt.title('Count for different Plateforms', fontsize=30)
plt.xlabel('Platform', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```

```python
plt.rcParams['figure.figsize']=(25,5)
sns.countplot(x=df['Year'])
plt.title('Count of game release in Years', fontsize=30)
plt.xlabel('Year', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```

```python
genre = df['Genre'].value_counts()
fig = px.pie(names=genre.index, values=genre, title='Genre Distribution')
fig.show()
```

```python
print('Total number of unique values present in column "Publisher" is: ', df['Publisher'].nunique())
```

```python
publisher = df['Publisher'].value_counts().head(10)
fig = px.bar(x=publisher.index, y=publisher, title='Top 10 Game Publisher')
fig.update_traces(marker_color='pink')
fig.show()
```

```python
na_sales = df[df['NA_Sales']!=0]

trace1 = go.Pie(labels=na_sales['Name'].value_counts().head(3).index, values=na_sales['Name'].value_counts().head(3),
               textinfo='percent',insidetextorientation='radial',legendgroup='show')

trace2 = go.Pie(labels=na_sales['Genre'].value_counts().head(3).index, values=na_sales['Genre'].value_counts().head(3),
               textinfo='percent',insidetextorientation='radial',legendgroup='show')

trace3 = go.Pie(labels=na_sales['Publisher'].value_counts().head(3).index, values=na_sales['Publisher'].value_counts().head(3),
               textinfo='percent',insidetextorientation='radial',legendgroup='show')

fig = make_subplots(rows = 3, cols=1, specs=[[{'type':'pie'}],
                                             [{'type':'pie'}],
                                             [{'type':'pie'}]],
                   subplot_titles=('Top 3 Games in NA','Top 3 Genre in NA','Top 3 Publisher in NA'))

fig.append_trace(trace1,1,1)
fig.append_trace(trace2,2,1)
fig.append_trace(trace3,3,1)

fig['layout'].update(title='Top 3 Sales In North America',
                    height=1100)

fig.show()
```

```python
eu_sales = df[df['EU_Sales']!=0]

trace1 = go.Pie(labels=eu_sales['Name'].value_counts().head(3).index, values=eu_sales['Name'].value_counts().head(3),
               textinfo='percent',insidetextorientation='radial',legendgroup='show')

trace2 = go.Pie(labels=eu_sales['Genre'].value_counts().head(3).index, values=eu_sales['Genre'].value_counts().head(3),
               textinfo='percent',insidetextorientation='radial',legendgroup='show')

trace3 = go.Pie(labels=eu_sales['Publisher'].value_counts().head(3).index, values=eu_sales['Publisher'].value_counts().head(3),
               textinfo='percent',insidetextorientation='radial',legendgroup='show')

fig = make_subplots(rows = 3, cols=1, specs=[[{'type':'pie'}],[{'type':'pie'}],[{'type':'pie'}]],
                   subplot_titles=('Top 3 Games in EU','Top 3 Genre in EU','Top 3 Publisher in EU'))

fig.append_trace(trace1,1,1)
fig.append_trace(trace2,2,1)
fig.append_trace(trace3,3,1)

fig['layout'].update(title='Top 3 Sales In Europe',
                    height=1100)

fig.show()
```

```python
jp_sales = df[df['JP_Sales']!=0]

trace1 = go.Pie(labels=jp_sales['Name'].value_counts().head(3).index, values=jp_sales['Name'].value_counts().head(3),
               textinfo='percent',insidetextorientation='radial',legendgroup='show')

trace2 = go.Pie(labels=jp_sales['Genre'].value_counts().head(3).index, values=jp_sales['Genre'].value_counts().head(3),
               textinfo='percent',insidetextorientation='radial',legendgroup='show')

trace3 = go.Pie(labels=jp_sales['Publisher'].value_counts().head(3).index, values=jp_sales['Publisher'].value_counts().head(3),
               textinfo='percent',insidetextorientation='radial',legendgroup='show')

fig = make_subplots(rows = 3, cols=1, specs=[[{'type':'pie'}],[{'type':'pie'}],[{'type':'pie'}]],
                   subplot_titles=('Top 3 Games in JP','Top 3 Genre in JP','Top 3 Publisher in JP'))

fig.append_trace(trace1,1,1)
fig.append_trace(trace2,2,1)
fig.append_trace(trace3,3,1)

fig['layout'].update(title='Top 3 Sales In Japan',
                    height=1100)

fig.show()
```
