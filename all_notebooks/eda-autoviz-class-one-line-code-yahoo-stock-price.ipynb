## Monthly Update EDA

```python
data = pd.read_csv('/kaggle/input/yahoo-finance-apple-inc-aapl/AAPL_Montly_updates.csv')
df = AV.AutoViz(filename="",sep=',', depVar='Adj Close', dfte=data, header=0, verbose=1, 
                 lowess=False, chart_format='svg', max_rows_analyzed=150000, max_cols_analyzed=30)
```
