```python
# Importing Pandas As pd
import pandas as pd

# Reading Data Frame
df =  pd.read_csv("../input/kotchandpurs-covid19-data-analysis/Covid Result Kcp.csv")
df.head(6)
```

```python
# Importing Seaborn
import seaborn as sns
#Sizing line plot
sns.set(rc={'figure.figsize':(20,8)})
# Making line Plot
sns.lineplot(data=df, x="Sample Tested", y="Found Positive")
```

```python
# Showing Sample Teated and Covid Positive Data By Barplot
sns.barplot(x="Sample Tested", y="Found Positive", data=df)
```

```python
# Showing Total Death Count By  Kdeplot
sns.kdeplot(data=df, x= "Total Death")
```

```python
# Importing Matplotlib for Color 
import matplotlib.pyplot as plt
# Setting Boxplot Showing Style
sns.set_theme(style="whitegrid")
# Making a Boxplot to Show Covid Positive
sns.boxplot(x=df["Found Positive"], color="g")
```

```python
# Showing The Sample test and covid Positive in Joint Plot
sns.jointplot(data = df, x="Sample Tested", y="Found Positive",color="r",  height=7)
```

```python
# Finding Co-relation 
cor = df.drop(['July'], axis=1).corr()
cor
# Creating Heatmap
plt.figure(figsize=(15,13))
sns.heatmap(cor, annot= True, cmap = plt.cm.CMRmap_r)
```
