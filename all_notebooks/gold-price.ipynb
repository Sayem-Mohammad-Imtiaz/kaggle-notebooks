```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data= pd.read_csv('/kaggle/input/yearly-gold-prices-from-1791-to-2020-in-usd/gold_price_1791_2020_yearly_denvergold.csv')
```

```python
data.head()
```

```python
data.corr()
```

```python
f,ax= plt.subplots(figsize=(5,5))
sns.heatmap(data.corr(), annot= True,  fmt='.1f' , ax=ax)
plt.show()
```

```python
data = data.rename(columns={'Nominal ounce closing price USD': 'Nominal'})
data = data.rename(columns={'Real ounce closing price USD': 'Real'})
```

```python
data.columns
```

```python
data.Nominal.plot(kind = 'line', color = 'g',label = 'Nominal ounce closing price USD',linewidth=3,alpha = 1,grid = True,linestyle = '-')

plt.legend(loc='upper right')     # legend = puts label into plot
plt.xlabel('time')              # label = name of label
plt.ylabel('Nominal ounce closing price USD')
plt.title('The time dependent change ')            # title = title of plot
plt.show()
```

```python
data.Real.plot(kind = 'line', color = 'r',label = 'Real ounce closing price USD',linewidth=3,alpha = 1,grid = True,linestyle = '-')

plt.legend(loc='upper right')     
plt.xlabel('time')              
plt.ylabel('Real ounce closing price USD')
plt.title('The time dependent change ')            
plt.show()
```

```python
data.plot(kind='scatter', color='r', alpha=1 , x='Real', y='Nominal')
plt.xlabel('Real ounce closing price USD')
plt.ylabel('Nominal ounce closing price USD')
plt.title('The comparison Real and Nominal closing price ')
plt.show()
```

```python
data.Real.plot(kind='hist', color='b', alpha=1, bins=35, figsize=(10,10))
plt.show()
```
