```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt  
%matplotlib inline

pd.set_option('display.max_columns', None)

salary_DF = pd.read_csv('../input/random-salary-data-of-employes-age-wise/Salary_Data.csv')
salary_DF.head()

salary_DF.shape
salary_DF.isnull().sum()
# check correlation
sns.heatmap(salary_DF.corr(), annot=True, cmap='viridis')
# Graph
salary_DF.plot(x='YearsExperience', y='Salary', style='o')
plt.title('Years of Experience vs Salary')
plt.xlabel('Years_of_experience')
plt.ylabel('Salary')
plt.show()

# Split the data
X = salary_DF['YearsExperience'].values.reshape(-1,1)
Y = salary_DF['Salary'].values.reshape(-1,1)

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.metrics import r2_score

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state=0)

regressor = LinearRegression()
regressor.fit(X_train, y_train) #training

# Intercept
regressor.intercept_

# Slope
regressor.coef_

y_pred = regressor.predict(X_test)
y_pred

final_DF = pd.DataFrame({'Actual' : y_test.flatten(), 'Predicted' : y_pred.flatten()})
final_DF
37731.0,40817.783270488115
122391.0,123188.08258899229
57081.0,65154.46261459161
63218.0,63282.410357352885
116969.0,115699.87356003735
109431.0,108211.66453108241
112635.0,116635.8996886567
55794.0,64218.43648597225
83088.0,76386.776158024

final_DF.shape

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('R-squared Error:', metrics.r2_score(y_test, y_pred))


# Graph
plt.scatter(X_test, y_test, color='green')
plt.plot(X_test, y_pred, color='blue', linewidth=2)
plt.show()

# to a csv file
final_DF.to_csv('predictions.csv',index=False)
```
