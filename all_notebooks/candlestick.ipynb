```python
# This Python 3 environment comes with many helpful analytics libraries installed
# It is defined by the kaggle/python Docker image: https://github.com/kaggle/docker-python
# For example, here's several helpful packages to load

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import mplfinance as mf
from sklearn.preprocessing import minmax_scale
```

```python
# original.csv may have more rows in reality, but we are only loading/previewing the first 1000 rows
data = pd.read_csv('/kaggle/input/5years-dailystock-quotes/original.csv', delimiter=',')
data.dataframeName = 'original.csv'
nRow, nCol = data.shape
print(f'There are {nRow} rows and {nCol} columns')
```

```python
data.head(5)
```

```python
#format date and set as index
new_format = "%Y-%m-%dT%H:%M:%SZ"
data['datetime'] = pd.to_datetime(data['datetime'], format=new_format)
data.rename(columns={'datetime':'Date'}, inplace=True)
data = data.set_index('Date')
#format rest of columns
data.rename(columns={'open_price':'Open'}, inplace=True)
data.rename(columns={'close_price':'Close'}, inplace=True)
data.rename(columns={'high_price':'High'}, inplace=True)
data.rename(columns={'low_price':'Low'}, inplace=True)
data.rename(columns={'volume':'Volume'}, inplace=True)
data.rename(columns={'symbol':'Symbol'}, inplace=True)
#data.rename(columns={'symbol':'Label'}, inplace=True)

data
```

```python
#choose ticker and subset
ticker = 'AAPL'
tick = data['Symbol'] == ticker
df = data[tick].loc[data.first_valid_index():]
df
```

```python
#scale volume
#df.loc[:,'Volume'] = minmax_scale(df['Volume'])
#df
```

```python
#trim to last 2 months 
row_split = df.shape[0] - 60
split = df.iloc[row_split+1:,:]
split.shape
```

```python
#explore types of candlesticks
mf.available_styles()
```

```python
mf.plot(split, 
        type = 'candle',
        style =  'yahoo',
        title = str(ticker) + ' walkthrough',
        ylabel = 'Prices',
        figratio=(18,6))
```

```python
#add volume :)

mf.plot(split, 
        type = 'candle',
        style =  'charles',
        volume = True )
```

```python
mc = mf.make_marketcolors(
                            up='tab:blue',down='tab:red',
                            wick={'up':'blue','down':'red'},
                            volume='tab:green',
                           )

s  = mf.make_mpf_style(marketcolors=mc)

mf.plot(split, 
        type = 'candle',
        style =  s,
        volume = True)
```

```python
mf.plot(split, 
        type = 'candle',
        style =  'yahoo',
        volume = True,
        show_nontrading = True,
        ylabel_lower='Volume of \n traded shares',
        axisoff=True,
        mav = 2)
```

```python
mf.plot(split, 
        type = 'candle',
        style =  'yahoo',
        figratio=(18,6),
        volume = True,
        show_nontrading = True,
        ylabel_lower='Volume of \n traded shares',
        axisoff=True,
        mav = (2,3,4))
```
