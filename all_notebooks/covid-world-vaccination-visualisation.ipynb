# Major contributor in vaccinations:********

```python
c_vac.groupby('vaccines')['country'].count().iplot(title='Covid vaccines Count')
```

```python
fig=px.choropleth(data_frame=c_vac,
                          locations=c_vac['country'],
                          locationmode='country names',
                          color=c_vac['vaccines'],
                          animation_group=c_vac['vaccines'])
fig.update_layout(dict1={'title':'Covid-Vaccines'})
fig.show() 
```

TOTAL VACCINATIONS PER HUNDRED:****

```python
c_vac.iplot(kind='bar',x='date',y='total_vaccinations_per_hundred',title='Date vs total_vaccinations_per_hundred')
```

```python
c_vac.iplot(kind='bar',x='country',y='total_vaccinations_per_hundred',title='total_vaccinations_per_hundred')
```

```python
fig=px.choropleth(data_frame=c_vac,
                          locations=c_vac['country'],
                          locationmode='country names',
                          color=c_vac['total_vaccinations_per_hundred'],
                          animation_frame=c_vac['date'],
                          animation_group=c_vac['vaccines'])
fig.update_layout(dict1={'title':'Total Vaccinations per hundred'})
fig.show() 
```

```python
fig = go.Figure()
for x in c_vac['country'].unique():
    fig.add_trace(go.Scatter(x=c_vac['date'], y=c_vac[c_vac['country']==x]['total_vaccinations_per_hundred'],
                         mode='lines+markers',
                         name=x))
fig.update_layout(dict1={'title':'Total Vaccinations Per Hundred'})
fig.show()
```
