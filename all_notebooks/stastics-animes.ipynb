```python
top1000 = df[:1000]
```

```python
top1000_genres = top1000['genres']
```

```python
split_genres = top1000_genres.str.split(',')
```

```python
genre_list = []
for genres in split_genres:
    for genre in genres:
        genre_list.append(genre)
```

```python
genre_sr = pd.Series(genre_list)
```

```python
genre_counts=genre_sr.value_counts()
```

```python
genre_counts.plot.barh(figsize=(10,10)).invert_yaxis()
```
