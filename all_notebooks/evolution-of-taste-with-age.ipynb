```python
#making some essential imports
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
dataFrame = pd.read_csv('/kaggle/input/have-you-ever-eaten-these-foods/responses.csv')
dataFrame = dataFrame.fillna(method = 'ffill')
dataFrame.info()
```

```python
####################################################################################
# indicator one correlartion : lets see how age correlates to food eaten           #
# for this we just need to map the bools and objects to values as follows          #
####################################################################################

def Mapper(featuresToMap):
    for colIds in featuresToMap:
        uvals = set(dataFrame[colIds])
        umap = {}
        i = 0
        
        for mem in uvals:
            umap[mem] = i
            i += 1
        print (colIds + '=>', end = '')
        print (umap)
        dataFrame[colIds] = dataFrame[colIds].map(umap)
```

```python
#lets map the required features
featuresNeedsMapping = ['Gender', 'Cockroach', 'Sushi',
       'Rocky Mountain oyster', 'Spam', 'Cow urine', 'Jalapeño',
       'Alligator meat', 'Paper', 'Hawaiian pizza', 'Pig\'s trotters', 'Durian',
       'Monkey brains', 'Frog legs', 'Caviar', 'Fugu', 'Balut', 'Lutefisk',
       'Century egg', 'Salo', 'Locusts']

Mapper(featuresNeedsMapping)
```

```python
### plotting correlation
plt.figure(figsize= (20, 10))
sb.heatmap(dataFrame.corr(), annot = True)
plt.show()
```

```python
#################################################
# some interesting observations                 #
#################################################
# As an individual grows old inclination towards#
# LuteFisk, hawaiian pizza, Sushi, increase     #
#                                               #
# whereas desire to eat century egg, fugu,      #
# cockraoch decreases with age                  #
#                                               #
# This might also be noticed tendecny           #
# to eat seafood increases with age             #
#################################################
```

```python
## defining a function for data visualization
def PlotDataGiveConclusion (featureVector):
    plt.figure(figsize= (30, 24))
    
    pltNum = 1
    for mem in featureVector:
        plt.subplot(3, 9 , pltNum)
        plt.grid(True)
        plt.title('Scatter Plot for '+ str(mem))
        sb.scatterplot(data = dataFrame, x = 'Age in years', y = mem)
        pltNum += 1
    
    plt.show()
```

```python
featuresToPlot = ['Cockroach', 'Sushi',
       'Rocky Mountain oyster', 'Spam', 'Cow urine', 'Jalapeño',
       'Alligator meat', 'Paper', 'Hawaiian pizza', 'Pig\'s trotters', 'Durian',
       'Monkey brains', 'Frog legs', 'Caviar', 'Fugu', 'Balut', 'Lutefisk',
       'Century egg', 'Salo', 'Locusts']

PlotDataGiveConclusion(featuresToPlot)
```

```python
## another Function for ploting regplot
def PlotDataGiveConclusion2 (featureVector):
    plt.figure(figsize= (30, 24))
    
    pltNum = 1
    for mem in featureVector:
        plt.subplot(4, 5 , pltNum)
        plt.grid(True)
        plt.title('Regplot Plot for '+ str(mem))
        sb.regplot(data = dataFrame, x = 'Age in years', y = mem , color = 'green')
        pltNum += 1
    
    plt.show()

    
PlotDataGiveConclusion2(featuresToPlot)
```
