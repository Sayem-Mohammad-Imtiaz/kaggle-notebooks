#### Loading Libraries...

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ta dando um warning na plotagem do pie plot
# investigar dpois
import warnings
warnings.filterwarnings("ignore")

sns.set_context("paper")
sns.set_style("darkgrid")
sns.set_palette('RdYlGn')
```

#### Starting analysis!

```python
df = pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
df.head()
```

#### Let's make some variable counts.

```python
plt.rcParams['figure.figsize'] = [15, 10]

fig, _ = plt.subplots(2,2)

sns.countplot(x='gender',
              data=df,
              ax=_[0][0])

sns.countplot(x='race/ethnicity',
              data=df,
              ax=_[0][1],
              order=['group A', 'group B', 'group C', 'group D', 'group E'])

sns.countplot(y='parental level of education',
              data=df,
              ax=_[1][0],
              order=['some college', 'some high school', 'high school', "associate's degree", "bachelor's degree", "master's degree"])

_[1][0].set_ylabel('')

sns.countplot(x='lunch',
              data=df,
              ax=_[1][1])

plt.suptitle('Variable Counts!')

plt.show()
```

<ul>
    <li> There are more women than men at school </li>
    <li> The predominant race/ethnicity is C, followed by D and B. A and E are a minority. </li>
    <li> Most student parents have little academic knowledge (some college) or just some specialization (associate's degree). </li>
    <li> Most use standard lunch. </li>
</ul>

#### We will now analyze academic performance by gender. What is the overall performance for each discipline?

```python
gender_math_scores = df.groupby('gender')['math score'].mean()
gender_reading_scores = df.groupby('gender')['reading score'].mean()
gender_writing_scores = df.groupby('gender')['writing score'].mean()
```

```python
plt.rcParams['figure.figsize'] = [16, 6]

fig, _ = plt.subplots(1,3)

sns.barplot(x=gender_math_scores.index,
            y=gender_math_scores.values,
            ax=_[0])

_[0].title.set_text('Math Scores by gender')

sns.barplot(x=gender_reading_scores.index,
            y=gender_reading_scores.values,
            ax=_[1])

_[1].title.set_text('Reading Scores by gender')

sns.barplot(x=gender_writing_scores.index,
            y=gender_writing_scores.values,
            ax=_[2])

_[2].title.set_text('Writing Scores by gender')

plt.show()
```

<ul>
    <li> Men have a higher average score than women in mathematics. </li>
    <li> Women have a higher average score than men in reading and writing. </li>
</ul>

#### Is there any relationship between the scores? That is, who has a good reading score, does well in writing?

```python
fig, _ = plt.subplots(1,3)

sns.scatterplot(x='reading score',
                y='writing score',
                data=df,
                alpha=0.6,
                ax=_[0])

_[0].title.set_text('Reading Score by Writing Score')

sns.scatterplot(x='reading score',
                y='math score',
                data=df,
                alpha=0.6,
                ax=_[1])

_[1].title.set_text('Reading Score by Math Score')

sns.scatterplot(x='writing score',
                y='math score',
                data=df,
                alpha=0.6,
                ax=_[2])

_[2].title.set_text('Writing Score by Math Score')

plt.show()
```

<ul>
    <li> Yep, that's right. Those who do well in reading, usually do well in writing too. </li>
    <li> But those who do well in reading and writing, generally do not do well in mathematics. </li>
</ul>

#### Let's ~scatter~ show this by gender!

```python
fig, _ = plt.subplots(1,3)

sns.scatterplot(x='reading score',
                y='writing score',
                hue='gender',
                data=df,
                alpha=0.6,
                ax=_[0])

_[0].title.set_text('Reading Score by Writing Score')

sns.scatterplot(x='reading score',
                y='math score',
                hue='gender',
                data=df,
                alpha=0.6,
                ax=_[1])

_[1].title.set_text('Reading Score by Math Score')

sns.scatterplot(x='writing score',
                y='math score',
                hue='gender',
                data=df,
                alpha=0.6,
                ax=_[2])

_[2].title.set_text('Writing Score by Math Score')

plt.show()
```

~this was a nice plot~
<ul>
    <li> Now we can see how the genders correlation between the scores! </li>
</ul>

#### Who did the test preparatory course do better or worse in the scores than those who didn't?

```python
fig, _ = plt.subplots(1,3)

sns.boxplot(x='test preparation course',
            y='writing score',
            data=df,
            ax=_[0])

_[0].title.set_text('Writing Score by those who studied or not')

sns.boxplot(x='test preparation course',
            y='reading score',
            data=df,
            ax=_[1])

_[1].title.set_text('Reading Score by those who studied or not')

sns.boxplot(x='test preparation course',
            y='writing score',
            data=df,
            ax=_[2])

_[2].title.set_text('Math Score by those who studied or not')

plt.show()
```

#### Does the academic level of the parents influence the students' average score?

Let's generate another column, called **general mean**. It will keep the overall average of each student (sum of the scores divided by 3).

```python
df['general mean'] = (df['math score'] + df['writing score'] + df['reading score']) / 3

df.head()
```

Now let's plot!

```python
plt.rcParams['figure.figsize'] = [15, 6]

fig, _ = plt.subplots(1,2)

sns.boxplot(x='parental level of education',
            y='general mean',
            data=df,
            order=['some college', 'some high school', 'high school', "associate's degree", "bachelor's degree", "master's degree"],
            ax=_[0])

_[0].tick_params('x', labelrotation=45)

df_grouped_median = df.groupby('parental level of education')['general mean'].median()

sns.pointplot(x=df_grouped_median.index,
              y=df_grouped_median.values,
              order=['some college', 'some high school', 'high school', "associate's degree", "bachelor's degree", "master's degree"],
              ax=_[1])

_[1].tick_params('x', labelrotation=45)

plt.suptitle("Does the academic level of the parents influence the students' average score?")

plt.show()
```

<ul>
    <li> Students with parents with master's degrees have higher average scores than the rest. </li>
    <li> Students with parents between some high school and high school are the ones with the worst averages in the scores. </li>
</ul>

#### Does the student's race / ethnicity influence his average score?

```python
fig, _ = plt.subplots(1,2)

sns.boxplot(x='race/ethnicity',
            y='general mean',
            data=df,
            order=['group A', 'group B', 'group C', 'group D', 'group E'],
            ax=_[0])

_[0].tick_params('x', labelrotation=45)

df_grouped_median = df.groupby('race/ethnicity')['general mean'].median()

sns.pointplot(x=df_grouped_median.index,
              y=df_grouped_median.values,
              order=['group A', 'group B', 'group C', 'group D', 'group E'],
              ax=_[1])

_[1].tick_params('x', labelrotation=45)

plt.suptitle("Does the student's race / ethnicity influence his average score?")

plt.show()
```

<ul>
    <li> Students of race/ethnicity in group E have better averages than the others. </li>
    <li> Students of race/ethnicity in group A have worse averages than the others. </li>
</ul>

#### Let's do the two previous plots by gender!

```python
fig, _ = plt.subplots(1,2)

gped_median_educ_gender = df.groupby(['parental level of education', 'gender'])['general mean'].median().reset_index()

sns.pointplot(x='parental level of education',
              y='general mean',
              hue='gender',
              data=gped_median_educ_gender,
              order=['some college', 'some high school', 'high school', "associate's degree", "bachelor's degree", "master's degree"],
              ax=_[0])

_[0].title.set_text("Does the academic level of the parents influence the students' average score? (by gender)")
_[0].tick_params('x', labelrotation=45)

gped_median_race_gender = df.groupby(['race/ethnicity', 'gender'])['general mean'].median().reset_index()

sns.pointplot(x='race/ethnicity',
              y='general mean',
              hue='gender',
              data=gped_median_race_gender,
              order=['group A', 'group B', 'group C', 'group D', 'group E'],
              ax=_[1])

_[1].title.set_text("Does the student's race / ethnicity influence his average score? (by gender)")
_[1].tick_params('x', labelrotation=45)

plt.show()
```

<ul>
    <li> Women are superior in scores in almost all scenarios. </li>
    <li> In the race/ethnicity of group E, the scores of both genders are very similar. </li>
</ul>

#### What percentage of students study before the exam in each ethnic group?

```python
gped_race_testprep = df.groupby('race/ethnicity')['test preparation course'].value_counts(normalize=True)
```

```python
plt.rcParams['figure.figsize'] = [15, 10]

f, _ = plt.subplots(2,3)

def plot_pie(gped_df, group, ax, f_n, s_n):
    
    gped_df.xs(group).plot(kind='pie', autopct='%1.0f%%', ax=ax[f_n][s_n])
    
    ax[f_n][s_n].title.set_text(group)

plot_pie(gped_race_testprep, 'group A', _, 0, 0)
plot_pie(gped_race_testprep, 'group B', _, 0, 1)
plot_pie(gped_race_testprep, 'group C', _, 0, 2)
plot_pie(gped_race_testprep, 'group D', _, 1, 0)
plot_pie(gped_race_testprep, 'group E', _, 1, 1)

_[1][2].set_visible(False)

plt.suptitle('What percentage of students study before the exam in each ethnic group?')

plt.show()
```

<ul>
    <li> Students from the ethnic group E are the most prepared for the exams. Because of this, their score is higher. </li>
    <li> The group that least prepares for the exams is group D. However, they have the second highest average of ethnic groups. </li>
</ul>

#### Students who have parents with little schooling usually prepare for exams?

```python
gped_educ_testprep = df.groupby('parental level of education')['test preparation course'].value_counts(normalize=True)
```

```python
f, _ = plt.subplots(2,3)

plot_pie(gped_educ_testprep, 'some college', _, 0, 0)
plot_pie(gped_educ_testprep, 'some high school', _, 0, 1)
plot_pie(gped_educ_testprep, 'high school', _, 0, 2)
plot_pie(gped_educ_testprep, "associate's degree", _, 1, 0)
plot_pie(gped_educ_testprep, "bachelor's degree", _, 1, 1)
plot_pie(gped_educ_testprep, "master's degree", _, 1, 2)

plt.suptitle('Students who have parents with little schooling usually prepare for exams?')

plt.show()
```

<ul>
    <li> Students who have parents with incomplete high school usually prepare for exams (more than the others). </li>
    <li> Students who have parents with bachelor's degree usually prepare for exams. </li>
    <li> Students who have parents with complete high school do not usually prepare for exams. </li>
</ul>

#### There's any correlation between parental level of education and race?

```python
gped_educ_race = df.groupby('parental level of education')['race/ethnicity'].value_counts(normalize=True).sort_index()
```

```python
f, _ = plt.subplots(2,3)

plot_pie(gped_educ_race, 'some college', _, 0, 0)
plot_pie(gped_educ_race, 'some high school', _, 0, 1)
plot_pie(gped_educ_race, 'high school', _, 0, 2)
plot_pie(gped_educ_race, "associate's degree", _, 1, 0)
plot_pie(gped_educ_race, "bachelor's degree", _, 1, 1)
plot_pie(gped_educ_race, "master's degree", _, 1, 2)

plt.suptitle("There's any correlation between parental level of education and race?")

plt.show()
```

<ul>
