```python
from IPython.display import Image
import pandas as pd
import numpy as np

from kaggle_secrets import UserSecretsClient

import pylast as lastfm
```

```python
user_secrets = UserSecretsClient()
lastfm_key = user_secrets.get_secret("lastfm_key")
lastfm_secret = user_secrets.get_secret("lastfm_secret")
network = lastfm.LastFMNetwork(api_key=lastfm_key, api_secret=lastfm_secret)
```

```python
df = pd.read_csv('/kaggle/input/muse-the-musical-sentiment-dataset/muse_v3.csv').astype({'valence_tags': np.float32, 'arousal_tags':np.float32, 'dominance_tags':np.float32})
```

```python
random_sample = df.sample(1)
track = random_sample['track'].to_list()[0]
artist = random_sample['artist'].to_list()[0]
print(f'artist: {artist}\ntrack: {track}')
```

```python

track = network.search_for_track(artist, track).get_next_page()[0]
```

```python
display(track)
```

```python
top_tags = track.get_top_tags(100)
tags = {tag_item.item.get_name():int(tag_item.weight) for tag_item in top_tags}
print(tags)
```
