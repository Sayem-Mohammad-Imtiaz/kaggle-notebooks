```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px
df = pd.read_csv('../input/gdp-annual-growth-for-each-country-1960-2020/GDP_annual_growth.csv').fillna(0)
%matplotlib inline
```

# ***GDP Growth Per Year***

```python
for i in range(1960,2020):
    print((' '*25)+"\033[1m"+'\033[94m'+'\033[4m'+f'{i}'+"\033[0m")
    fig = px.choropleth(df, locations="Country Code",color=df[f'{i}'],hover_name="Country Name",color_continuous_scale=px.colors.sequential.Rainbow)
    fig.show()
```

# **Average GDP Change from 1960-2019**

```python
for i in range(1960,2020):
    print("\033[1m"+'\033[94m'+'\033[4m'+f"{i}"+"\033[0m"+'\033[94m'+f" Average Change in gdp is {df.groupby('Country Name').mean()[str(i)].mean()}"+"\033[0m")
```
