```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('../input/palestine-body-count/Palestine Body Count.csv')
data.shape
```

# 2. Data Preprocessing

```python
data.info()
```

We have some problems here >> missing values and most of columns are integer but represent as object.

```python
data.isnull().sum()
```

```python
Total = data.isnull().sum().sort_values(ascending=False)
percent = (data.isnull().sum() / data.isnull().count()).sort_values(ascending=False)
missing_data = pd.concat([Total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(25)
```

The Missing from the injuries of the Israelis is 47%

```python
data.fillna(0,inplace = True)
```

```python
data.isnull().sum()
```

```python
data
```

Coulmns have 'inc.Jun' and 'incl.Aug', so we must replace them.

```python
data['Palestinians Injuries']=data['Palestinians Injuries'].replace("(incl. Jun)", np.nan)
data['Israelis Injuries']=data['Israelis Injuries'].replace("(incl. Jun)", np.nan)
data['Palestinians Injuries']=data['Palestinians Injuries'].replace("(incl. Aug)", np.nan)
data['Israelis Injuries']=data['Israelis Injuries'].replace("(incl. Aug)", np.nan)
```

```python
data.fillna(0,inplace=True)
```

```python
data.info()
```

##### I see the last 2 row don't have a year and month so i decided to take rows above them.

```python
df= data.iloc[0:249]
df
```

##### Convert the object columns to int

```python
df = df.astype({'Year': int,"Palestinians Injuries":int,"Israelis Injuries":int,"Palestinians Killed":int,"Israelis Killed":int})
```

```python
df.info()
```

```python
df.describe()
```

# 3. EDA

```python
import seaborn as sns
sns.pairplot(df,hue='Month')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
sns.heatmap(df.corr(),annot=True, cmap='RdYlGn')
```

#### Note that there is a correlation between the number of Israelis Injuries and Palestinians Injuries


```python
import plotly.express as px
pl=df[{'Year','Palestinians Injuries','Year'}]
gra=px.line(pl,x='Year',y='Palestinians Injuries',color='Year',title='Palestinians Injuries',range_x=[1999,2022])
gra.show()

isr=df[{'Year','Israelis Injuries'}]
gra=px.line(df,x='Year',y='Israelis Injuries',color='Year',title='Israelis Injuries',range_x=[1999,2022])

gra.show()
```

```python
plt.rc('font', size=12)
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(df['Year'], df['Palestinians Killed'], color='tab:red', label='Palestinians Killed')
ax.set_xlabel('Years')
ax.set_ylabel('Count')
ax.set_title('Palestinians Killed')
ax.grid(True)
ax.legend(loc='upper left')

plt.rc('font', size=12)
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(df['Year'], df['Israelis Killed'], color='tab:blue', label='Israelis Killed')
# Same as above
ax.set_xlabel('Years')
ax.set_ylabel('Count')
ax.set_title('Israelis Killed')
ax.grid(True)
ax.legend(loc='upper left')

```

```python
plt.figure(figsize=(12,6))
plt.bar(df['Year'],df['Palestinians Killed'] , label = "Palestinians Killed" ) 
plt.bar(df['Year'],df['Israelis Killed'] , label = "Israelis Killed")
plt.xlabel('Years')
plt.legend()
plt.show()

```
