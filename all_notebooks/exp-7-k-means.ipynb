```python
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
data = pd.read_csv('../input/customer-segmentation-tutorial-in-python/Mall_Customers.csv').sample(frac=1).reset_index(drop=True)
data.head()
```

```python
col_1 = 3
col_2 = 4
X = data.iloc[:,[col_1,col_2]].to_numpy()
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i,init ='k-means++')
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('No. of Clusters')
plt.xlabel('WCSS')
plt.show()
```

```python
cno = 4
kmeans = KMeans(n_clusters = cno,init ='k-means++')
y_kmeans = kmeans.fit_predict(X)
```

```python
data['clusters'] = y_kmeans
data.head()
```

```python
df = []
for i in range(cno):
    df.append(data[data.clusters==i])
    
colors = ['red','green','black','blue','orange','cyan','grey']
for i in range(cno):
    plt.scatter(df[i].iloc[:,col_1],df[i].iloc[:,col_2],color=colors[i],label='cluster '+str(i+1))

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=150,c='yellow',label='Centroids')
plt.title('Cluster of Customers')
plt.xlabel(data.columns[col_1])
plt.ylabel(data.columns[col_2])
plt.legend()
plt.show()
```
