```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../input/gaussian-mixture-models/Clustering_gmm.csv")
df=df.dropna()
df.head()
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['Weight'],df['Height'])
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title("Data Distribution")
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
gm = GaussianMixture(n_components=4)
gm.fit(df)
```

```python
labels = gm.predict(df)
frame = pd.DataFrame(df)
frame['Clusters'] = labels
```

```python
frame.columns = ['weight', 'height', 'Clusters']
color = ['blue', 'green', 'cyan', 'black']
for k in range (0, 4):
    df = frame[frame['Clusters']==k]
    plt.scatter(df['weight'],df['height'],c=color[k])
plt.show()
```

# **PART 2 IRIS DATASET**

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn import datasets
import seaborn as sns
sns.set(style="white", color_codes=True)
# show plots inline
%matplotlib inline
```

```python
#loading datasets
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)
df.head()
```

```python
X = df.iloc[:,0:4]
y = df.iloc[:,-1]
print(X.sample(5))
print(y.sample(5))
```

```python
#scatterplot
sns.FacetGrid(df,hue='species',size=4).map(plt.scatter,"sepal length (cm)","sepal width (cm)").add_legend();
```

```python
sns.FacetGrid(df,hue='species',size=4).map(plt.scatter,"petal length (cm)","petal width (cm)").add_legend();
```

```python
sns.swarmplot(x="species",y="petal length (cm)",data=df)
plt.grid()
```

```python
#Scaling the dataset
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
scaler.fit(X)
X_scaled_array = scaler.transform(X)
X_scaled = pd.DataFrame(X_scaled_array, columns = X.columns)
X_scaled.sample(5)
```

```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=3, covariance_type='full')
gmm.fit(X_scaled)
# predict the cluster for each data point
y_cluster_gmm = gmm.predict(X_scaled)
y_cluster_gmm
```

```python
from sklearn.metrics.cluster import adjusted_rand_score
score = adjusted_rand_score(y, y_cluster_gmm)
score
```
