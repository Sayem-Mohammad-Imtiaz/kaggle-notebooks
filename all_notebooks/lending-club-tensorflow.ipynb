```python
def fill_mort_acc(total_acc,mort_acc):

    if np.isnan(mort_acc):
        return total_acc_avg[total_acc]
    else:
        return mort_acc
```

```python
x=df.drop('Fully Paid',axis=1).values
y=df['Fully Paid'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit_transform(x_train)
```

```python
scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
```

```python
model=Sequential()

model.add(Dense(50,activation='softplus'))
model.add(Dropout(0.1))


model.add(Dense(25,activation='softplus'))
model.add(Dropout(0.1))


model.add(Dense(13,activation='softplus'))
model.add(Dropout(0.1))


model.add(Dense(units=1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='adam')

```

```python
model.fit(x=x_train,y=y_train,epochs=25,
         batch_size=256,validation_data=(x_test,y_test))
```

```python
losses=pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
predictions=model.predict_classes(x_test)
```

```python
print(classification_report(y_test,predictions))
```

```python
confusion_matrix(y_test,predictions)
```
