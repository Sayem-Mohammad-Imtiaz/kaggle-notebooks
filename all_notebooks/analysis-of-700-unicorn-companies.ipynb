# Content

List of 700+ Unicorn Companies across the globe.

Well, Unicorn company? A company which has a valuation of over $1 billion.

A unicorn company, or unicorn startup, is a private company with a valuation over $1 billion. As of June 2021, there are more than 700 unicorns around the world.

New subsets include a decacorn, companies which are valued at over $10 billion, and a hectocorn, companies which are valued at over $100 billion.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings 
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("../input/list-of-700-unicorn-companies-across-the-globe/List of 700 Unicorn Companies.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Valuation (in $B)"] = df["Valuation (in $B)"].apply(lambda x : x.replace("$","")) 
```

```python
df["Valuation (in $B)"] = df["Valuation (in $B)"].astype("float")
```

```python
df["Valuation (in $B)"].describe()
```

# EDA

## Country of origin

### Counts of Country of origin

```python
df["Country of origin"].value_counts()
```

```python
# Top 10
plt.figure(figsize=(12,6))
sns.countplot(data=df,x="Country of origin", order = df["Country of origin"].value_counts().iloc[:10].index)
plt.show()
```

### Valuation of Country of origin

```python
top_10 = pd.DataFrame(df.groupby(['Country of origin'])['Valuation (in $B)'].sum().sort_values(ascending=False)[:10])
top_10
```

```python
plt.figure(figsize=(12,7))
sns.barplot(data=top_10, x=top_10.index, y="Valuation (in $B)")

plt.show()
```

## Industries 

### Counts of Industries

```python
df["Industry"].value_counts()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(data=df,y="Industry", order = df["Industry"].value_counts()[:10].index)
plt.show()
```

### Valuation of Industries

```python
top_10 = pd.DataFrame(df.groupby(['Industry'])['Valuation (in $B)'].sum().sort_values(ascending=False)[:10])
top_10
```

```python
plt.figure(figsize=(15,7))
sns.barplot(data=top_10, x="Valuation (in $B)", y=top_10.index)
plt.show()
```

## City

### Counts of City

```python
df["City"].value_counts()[:10]
```

```python
plt.figure(figsize=(12,8))
sns.countplot(data=df,x="City", order=df["City"].value_counts()[:10].index)
plt.grid(axis="y",color="r", linestyle="--")
plt.show()
```

### Valuation of City

```python
top_10 = pd.DataFrame(df.groupby(['City'])['Valuation (in $B)'].sum().sort_values(ascending=False)[:10])
top_10
```

```python
plt.figure(figsize=(12,7))
sns.barplot(data=top_10, x=top_10.index, y="Valuation (in $B)")

plt.show()
```
