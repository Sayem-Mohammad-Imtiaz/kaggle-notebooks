```python
questions_df = pd.read_csv("/kaggle/input/pythonquestions/Questions.csv", encoding="ISO-8859-1")
questions_df.head(5)
```

```python
def count_word_in_questions(word, column):
    return sum([1 if word.lower() in item.lower() else 0 for item in questions_df[column]])
```

```python
count_word_in_questions('flask', 'Title') + count_word_in_questions('flask', 'Body')
```

```python
count_word_in_questions('fastapi', 'Title') + count_word_in_questions('fastapi', 'Body') + count_word_in_questions('fast api', 'Title') + count_word_in_questions('fast api', 'Body')
```
