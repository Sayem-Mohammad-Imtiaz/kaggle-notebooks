```python
# import neccessary modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
df1 = pd.read_csv("../input/inshort-news/inshort_news_data-1.csv", index_col=0)
df2 = pd.read_csv("../input/inshort-news/inshort_news_data-2.csv", index_col=0)
df3 = pd.read_csv("../input/inshort-news/inshort_news_data-3.csv", index_col=0)
df4 = pd.read_csv("../input/inshort-news/inshort_news_data-4.csv", index_col=0)
```

```python
# combining into single dataframe
df = pd.concat([df1, df2, df3, df4], ignore_index=True)
```
