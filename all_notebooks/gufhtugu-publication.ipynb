```python
df = pd.read_csv("/kaggle/input/gufhtugu-publications-dataset-challenge/GP Orders - 5.csv")
df.sample(10)
```

Rename column names for better EDA

```python
df = df.rename(columns={'Order Number': 'order_number',"Order Status":"order_status", "Book Name":"book_name","Order Date & Time":"order_date","City":"billing_city", "Payment Method":"payment_method", "Total items":"total_books", "Total weight (grams)":"grams"})
df.head()
```

Check Null values in dataset

```python
df.isnull().sum().sort_values(ascending = False)
```

Drop NaN Values

```python
df.dropna(inplace=True)
df.isnull().sum()
```

Order Status Frequency

```python
df.order_status.value_counts()
```

```python
plt.figure(figsize=(12,6))
bar= df.order_status.value_counts().plot(kind="bar")
plt.xlabel("Order Status")
plt.ylabel("Total orders")
```

> Payment Methods

```python
complete_df = df.loc[(df.order_status == 'Completed')]
complete_df.payment_method.value_counts()
```

```python
plt.figure(figsize=(12,6))
histogram = complete_df.payment_method.hist()
plt.xlabel("Payment Method")
plt.ylabel("Number of Completed Orders")
```

Seprate the books in each order on "/"

```python
df = df.assign(book_name=df.book_name.str.split("/")).explode("book_name")
df.book_name.value_counts()[:5]
```

Best 5 Selling Books

```python
plt.figure(figsize=(14,5))
top_5_seller = df.book_name.explode().value_counts()[:5].plot.bar()
plt.title("top_5_Selling_Books")
plt.xlabel("Books")
plt.ylabel("Number of Orders")
```

Least 5 Selling Books

```python
plt.figure(figsize=(14,5))
least_5_selleing_books = df.book_name.explode().value_counts()[-5:].plot.bar()

plt.title("Least 5 Selling Books")
plt.xlabel("Books")
plt.ylabel("Numbers of Orders")
```

Total Sold BOOKS

```python
total_sold = df.book_name.explode().value_counts().sum()
print(f"Total Books Sold by Gufhtugu Publications are {total_sold}")
```

Find a correlation between date and time with order statusÂ¶

```python
df['order_date']= pd.to_datetime(df['order_date'])
df['year'] = df['order_date'].apply(lambda x : x.year)
df['month'] = df['order_date'].apply(lambda x : x.month)
df['day'] = df['order_date'].apply(lambda x : x.day_name())
df['weekday'] = df['order_date'].apply(lambda x : x.weekday())
df_new=df[['order_number', 'order_status', 'book_name', 'order_date', 'billing_city', 'year', 'month', 'day','weekday']]
df.head()
```

```python
years= df.year.value_counts()
years
```

```python
plt.figure(figsize=(14,5))
years.plot(kind="bar")
plt.xlabel("years")
plt.ylabel("Number of Sales")
```

```python
daily_sales = df.day.value_counts()
daily_sales
```

```python
plt.figure(figsize=(14,6))
daily_sales.plot(kind="bar")
plt.xlabel("days")
plt.ylabel("Number of Sales")
```

```python
Monthly_sale = df.month.value_counts()
Monthly_sale
```

```python
plt.figure(figsize=(14,6))
Monthly_sale.plot(kind="bar")
plt.xlabel("Month")
plt.ylabel("Number of Sales")
```
