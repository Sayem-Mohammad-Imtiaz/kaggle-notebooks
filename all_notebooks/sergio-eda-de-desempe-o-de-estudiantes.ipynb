```python
# Arriba  -  Analsis por generos  y raza

import plotly.express as px
import pandas as pd

#data = pd.read_csv("StudentsPerformance.csv")


fig = px.sunburst(data,path=['gender','race/ethnicity','parental level of education'],
                  values='math score',color='race/ethnicity',)
#hover_data = ['Sales','Quantity',Profit])
fig.update_layout(height=1000,title_text='Analsis de Generos')
fig.show()
```

```python
#  Analsis por grupo E

import plotly.express as px
import pandas as pd

#data = pd.read_csv("StudentsPerformance.csv")

df_grupoE = data[data['race/ethnicity'] == 'group E']

fig = px.sunburst(df_grupoE,path=['gender','race/ethnicity','parental level of education'],
                  values='math score',color='parental level of education',)
#hover_data = ['Sales','Quantity',Profit])
fig.update_layout(height=1000,title_text='Analsis de Grupo E')
fig.show()
```

```python
#Ordenamos los valores por columna math score y gender en orden ascendente
data = data.sort_values(['math score','gender'],ascending=True)
data
```

```python
#Analisis de Calificaciones

mathc = data.groupby('gender').sum()['math score'].reset_index()
readc = data.groupby('gender').sum()['reading score'].reset_index()
wrtc = data.groupby('gender').sum()['writing score'].reset_index()

```

```python
# Grafica utilizando Barras

trace1 = go.Bar(x = mathc['gender'],
              y = mathc['math score'],
              name = 'math score',
              marker_color = 'blue')

trace2 = go.Bar(x = readc['gender'],
              y = readc['reading score'],
              name = 'reading score',
              marker_color = 'red')

trace3 = go.Bar(x = wrtc['gender'],
              y = wrtc['writing score'],
              name = 'writing score',
              marker_color = 'green')

layout = go.Layout(barmode = 'group',
                  bargap = 0,
                  bargroupgap = 0,
                  title = "Calificaciones (Bar Chart)",
                  xaxis = dict(title='gender',zeroline= False,
                         gridcolor='rgb(183,183,183)',showline=True),
                  yaxis = dict(title='math score',zeroline= False,
                            gridcolor='rgb(183,183,183)',showline=True),
                  font = dict(family='Courier New, monospace', size=12, color='rgb(0,0,0)'),
                  legend=dict(x=0, y=1.0, bgcolor='rgba(255, 255, 255, 0)', bordercolor='rgba(255, 255, 255, 0)'),
)

data = [trace1, trace2, trace3]
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

```python
# Grafica utilizando Lineas y marcadores

trace1 = go.Scatter(x = mathc['gender'],
              y = mathc['math score'],
              name = 'math score',
              mode='lines+markers',
              line=dict(color='blue', width=1))

trace2 = go.Scatter(x = readc['gender'],
              y = readc['reading score'],
              name = 'reading score',
             mode='lines+markers',
              line=dict(color='red', width=1))

trace3 = go.Scatter(x = wrtc['gender'],
              y = wrtc['writing score'],
              name = 'writing score',
              mode='lines+markers',
              line=dict(color='green', width=1))

layout = go.Layout(barmode = 'group',
                  bargap = 0,
                  bargroupgap = 0,
                  title = "Calificaciones (Line Chart)",
                  xaxis = dict(title='gender',zeroline= False,
                         gridcolor='rgb(183,183,183)',showline=True),
                  yaxis = dict(title='math score',zeroline= False,
                            gridcolor='rgb(183,183,183)',showline=True),
                  font = dict(family='Courier New, monospace', size=12, color='rgb(0,0,0)'),
                  legend=dict(x=0, y=1.0, bgcolor='rgba(255, 255, 255, 0)', bordercolor='rgba(255, 255, 255, 0)'),
)

data = [trace1, trace2, trace3]
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```
