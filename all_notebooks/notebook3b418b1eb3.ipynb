We start off by reading the csv file, and importing some important libraries we need to visualise the data:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris = pd.read_csv("/kaggle/input/iris-flower-dataset/IRIS.csv")
```

After displaying some basic information of the file given, we can start to visualise the information by using some important python libraries

First, we start off by showing the histogram from the pandas library


```python
iris.hist(bins = 40, figsize=(12,8))
```

Visualising using Scatter

```python
ig, sc = plt.subplots()
sc.scatter(iris['sepal_length'], iris['sepal_width'], c = "green")
```

```python
ig2, sc2 = plt.subplots()
sc2.scatter(iris['petal_length'], iris['petal_width'], c = "red")
```

Now to do the count plot for each piece of data provided:

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(20,4))
sns.countplot(x=iris['sepal_length'],data=iris,hue = 'species',palette='bright')
plt.show()
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(20,4))
sns.countplot(x=iris['sepal_width'],data=iris,hue = 'species',palette='bright')
plt.show()
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(20,4))
sns.countplot(x=iris['petal_length'],data=iris,hue = 'species',palette='bright')
plt.show()
```

```python
sns.set_theme(style="darkgrid")
plt.figure(figsize=(20,4))
sns.countplot(x=iris['petal_width'],data=iris,hue = 'species',palette='bright')
plt.show()
```

In order to visualise the Pie chart of the data, I defined labels for the pie chart & included the values the sizes variable

```python
labels = 'iris setosa', 'iris versicolor', 'iris virginica'
sizes = iris.groupby('species')['sepal_length'].mean().values

plt.pie(sizes,labels=labels, autopct ='%1.1f%%', shadow=True, startangle=90)
plt.title('Sepal length per species')
plt.show()

labels = 'iris setosa', 'iris versicolor', 'iris virginica'
sizes = iris.groupby('species')['sepal_width'].mean().values

plt.pie(sizes,labels=labels, autopct ='%1.1f%%', shadow=True, startangle=90)
plt.title('Sepal width per species')
plt.show()

labels = 'iris setosa', 'iris versicolor', 'iris virginica'
sizes = iris.groupby('species')['petal_length'].mean().values

plt.pie(sizes,labels=labels, autopct ='%1.1f%%', shadow=True, startangle=90)
plt.title('petal length per species')
plt.show()

labels = 'iris setosa', 'iris versicolor', 'iris virginica'
sizes = iris.groupby('species')['petal_width'].mean().values

plt.pie(sizes,labels=labels, autopct ='%1.1f%%', shadow=True, startangle=90)
plt.title('petal width per species')
plt.show()
```

Since we have 3 species we can use each trait such as length & width for each specimen in the pie chart
