```python
# SVC ON 'heart_disease' DATASET TO CALCULATE ACCURACY AND COMPARE THE PREDICTED AND ORIGINAL VALUES. 
# ACCURACY STARTS DECREASING WHEN RELATABLE COLUMN IS REMOVED FROM COLUMN UNDER CONSIDERATION.

import pandas as pd
from sklearn.svm import SVC
from sklearn import model_selection

df=pd.read_csv('../input/heart-disease-patients-details/heart_disease.csv')

Y=df['result']
X=df.loc[:,'age':'oldpeak']

X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,random_state=7,test_size=0.33)

sc=SVC(kernel='poly',random_state=100) 
sc.fit(X_train,Y_train)

print(sc.score(X_train,Y_train))

prediction=sc.predict(X_test)
print(prediction)
print(Y_test)
```
