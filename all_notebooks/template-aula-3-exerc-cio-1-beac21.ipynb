# Questão 1

Construa um gráfico de violino que compara o número de eleitores do gênero masculino e gênero feminino das 15 cidades com o maior número de eleitores do estado da Bahia.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../input/BR_eleitorado_2016_municipio.csv', delimiter=',')
df.head()
```

```python
ba = df[df['uf']=='BA'].nlargest(columns='total_eleitores',n=15)
ba
```

```python

plt.violinplot([ba['gen_masculino'],ba["gen_feminino"]],showmeans=True, showmedians=True,widths = 0.5, bw_method=0.5) #default
plt.xticks([0,1,2], ('', 'Masculino','Feminino'))

plt.show()
```

# Questão 2

Construa um segundo gráfico de violino que compara o número de eleitores com idades entre 18-20 anos, 21-24 anos, 70-79 anos e eleitores com mais de 79 anos, para as 15 cidades com o maior número de eleitores do estado de Minas Gerais.

```python
mg = df[df['uf']=='MG'].nlargest(columns='total_eleitores',n=15)
mg
```

```python
plt.violinplot([mg['f_18_20'],mg["f_21_24"],mg["f_70_79"],mg["f_sup_79"]],showmeans=True, showmedians=True,widths = 0.5, bw_method=0.5) #default
plt.xticks([0,1,2,3,4], ('', '18 a 20','21 a 24','70 a 79','superior a 79'))

plt.show()
```
