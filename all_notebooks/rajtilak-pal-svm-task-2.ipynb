```python
df=pd.read_csv('/kaggle/input/pima-indians-diabetes-database/diabetes.csv')
```

```python
df
```

```python
X=df.drop(columns='Outcome')
y=df['Outcome']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

### Applying a Simple SVM to see initially how the accuracy scores are

```python
svm=SVC()
```

```python
svm.fit(X_train,y_train)
```

```python
y_pred_train=svm.predict(X_train)
y_pred_test=svm.predict(X_test)
```

```python
print('Training Accuracy: ',accuracy_score(y_pred_train,y_train))
print('Testing Accuracy: ',accuracy_score(y_pred_test,y_test))
```

```python
print("Training Classification Report: \n", classification_report(y_train,y_pred_train))
print("Testing Classification Report: \n", classification_report(y_test,y_pred_test))
```

### Finding best C score for Linear Kernel

```python
c=np.linspace(0.05,20,20)

scores_train=[]
scores_test=[]

for i in range(len(c)):
    svm=SVC(kernel='linear',C=c[i])
    svm.fit(X_train,y_train)
    y_pred_train=svm.predict(X_train)
    y_pred_test=svm.predict(X_test)
    scores_train.append(accuracy_score(y_pred_train,y_train))
    scores_test.append(accuracy_score(y_pred_test,y_test))
    print(f"Completed {i}")
```

```python
plt.plot(c,scores_train, label='Training Accuracies')
plt.plot(c,scores_test, label='Testing Accuracies')
plt.legend()
```

```python
print(max(scores_test))
c[np.argmax(scores_test)]
```

#### Best Accuracy for Linear kernel 78.57% was at C=1.1
