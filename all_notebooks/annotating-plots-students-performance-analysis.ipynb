**Students Performance in Exam**

Aim:

1. Showing the scores of math and reading by adhering to writing scores. (Scores will be above the average of the students.)
2. Creating the histogram both for female and male.


```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
```

```python
df=pd.read_csv("../input/students-performance-in-exams/StudentsPerformance.csv")
```

```python
df.head(5)
```

```python
df.isna().sum()
df = df.drop(columns = ["race/ethnicity", "test preparation course"])
```

```python
pd.unique(df["lunch"])
```

```python
df.describe()
```

```python
df_female = df[df["gender"] == "female"]
df_male = df[df["gender"] == "male"]

df_female.head(5)
df_male.head(5)
```

# **Overview** 

```python
import plotly.figure_factory as ff
import numpy as np

# Add histogram data
x1 = df_female["math score"]
x2 = df_female["reading score"]
x3 = df_male["math score"]
x4 = df_male["reading score"]

# Group data together
hist_data = [x1, x2, x3, x4]

group_labels = ['Female Math Score', 'Female Reading Score', "Male Math Score", "Male Reading Score"]

# Create distplot with custom bin_size
fig = ff.create_distplot(hist_data, group_labels, bin_size=.2)
fig.show()
```

```python
df_female.mean()
```

```python
df_male.mean()
```

**Taking the values that are above average.**

```python
abv_f = df_female[(df_female["math score"] > df_female["math score"].mean()) & (df_female["reading score"] > df_female["reading score"].mean()) & (df_female["writing score"] > df_female["writing score"].mean()) ]
print(abv_f)
abv_m = df_male[(df_male["math score"] > df_male["math score"].mean()) & (df_male["reading score"] > df_male["reading score"].mean()) & (df_male["writing score"] > df_male["writing score"].mean()) ]
```

**Importing some libraries from plotly.**

```python
from matplotlib.patches import Ellipse
from matplotlib.text import OffsetFrom
import plotly.express as px
```

```python
# plot >> gender distribution above avg .

# important_features = ['parental level of education', 'lunch', 'math score', 'reading score', 'writing score']
# df_pl = abv_f[important_features]


#.......main plot (scatter)
fig = px.scatter(abv_f,
                 x='reading score',
                 y='math score',
                 color="writing score",
                
                 width=750,
                 height=400)

fig2 = px.scatter(df,
                 x='reading score',
                 y='math score',
                 color="gender",
                
                 width=750,
                 height=400)
               
#......annotation (box)
fig.add_shape(type="rect",
              xref="x", yref="y",
              x0=80, x1=100, y0=80, y1=100,
              fillcolor="lightgray",    
              line_color="yellow",
              opacity=0.25
              )
fig2.add_shape(type="rect",
              xref="x", yref="y",
              x0=85, x1=100, y0=85, y1=100,
              fillcolor="red",    
              line_color="red",
              opacity=0.25
              )
#......update the plot as you wish
fig.update_layout(title='<b>Math & Reading & Writing Scores<b>',
                  titlefont={'size': 24},
                  template='plotly_dark'
                  )
fig.show()
fig2.show()
```
