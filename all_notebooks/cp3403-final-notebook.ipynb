# Preprocessing for Association Rule Mining

```python
def round_down(num, divisor):
    return num - (num % divisor)


def processMean(df):
    for i in df["MEAN"]:
        if(math.isnan(i) == False):
            y = "{}-{}".format(round_down(i, 10), round_down(i, 10)+10)
            df["MEAN"].replace({i: y}, inplace=True)


def processPercent(df):
    for i in df["ABOVE NATIONAL MINIMAL STANDARD %"]:
        if(math.isnan(i) == False):
            if (70 > i):
                df["ABOVE NATIONAL MINIMAL STANDARD %"].replace(
                    {i: "Below 70%"}, inplace=True)
            else:
                y = "{}-{}".format(round_down(i, 5), round_down(i, 5)+5)
                df["ABOVE NATIONAL MINIMAL STANDARD %"].replace(
                    {i: y}, inplace=True)


df_grammar = pd.read_csv("../input/australianeducationdata/Results2019Year9grammarandpunctuationgeolocation.csv")
df_numeracy = pd.read_csv("../input/australianeducationdata/Results2019Year9numeracygeolocation.csv")
df_reading = pd.read_csv("../input/australianeducationdata/Results2019Year9readinggeolocation.csv")
df_spelling = pd.read_csv("../input/australianeducationdata/Results2019Year9spellinggeolocation.csv")
df_writing = pd.read_csv("../input/australianeducationdata/Results2019Year9writinggeolocation.csv")


processMean(df_grammar)
processPercent(df_grammar)

processMean(df_numeracy)
processPercent(df_numeracy)

processMean(df_reading)
processPercent(df_reading)

processMean(df_spelling)
processPercent(df_spelling)

processMean(df_writing)
processPercent(df_writing)


df_grammar.to_csv('df_grammar.csv', index=False)
df_numeracy.to_csv('df_numeracy.csv', index=False)
df_reading.to_csv('df_reading.csv', index=False)
df_spelling.to_csv('df_spelling.csv', index=False)
df_writing.to_csv('df_writing.csv', index=False)
```
