```python
import numpy as np
import pandas as pd
import os
import folium
from folium import plugins
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data0=pd.read_csv('../input/fast-food-restaurants-across-us/Fast_Food_Restaurants_US.csv')
data0
```

```python
print(data0.columns.tolist())
```

### Fast Food Restaurant Ranknig

```python
data2a=data0[['name']]
data2a['number']=1
data2a=data2a.groupby('name',as_index=False).sum()
data2a=data2a.sort_values('number',ascending=False)
data2a
```

```python
fig = px.bar(data2a[0:30], x='name', y='number',title="Fast Food Restaurant Ranknig")
fig.show()
```

### Subway Ranknig by City

```python
data2b=data0[data0['name']=='Subway'][['city']]
data2b['number']=1
data2b=data2b.groupby('city',as_index=False).sum()
data2b=data2b.sort_values('number',ascending=False)
data2b
```

```python
fig = px.bar(data2b[0:30], x='city', y='number',title="Subway Ranknig by City")
fig.show()
```

### Location of Subway in Chicago 

```python
data3=data0[data0['name']=='Subway'][data0['city']=='Chicago'][['latitude','longitude']]
data3=data3.fillna('N')
data3
```

```python
eq_map = folium.Map(location=[41.9,-87.7],tiles='Stamen Terrain',zoom_start=10.0,min_zoom=2.0)
eq_map.add_child(plugins.HeatMap(data3))
eq_map
```
