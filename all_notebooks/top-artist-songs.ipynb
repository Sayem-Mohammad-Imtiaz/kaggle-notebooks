```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/kaggle/input/billboard-the-hot-100-songs/charts.csv")
df.head(5)
```

*Top 10 artists plus mentions in The Top 100*****

```python
top = 10

df2 = df['artist']
artist_frec = {'Artist':list(df2.value_counts().index),
               'Frecuency':list(df2.value_counts().values)}
df_artistTop = pd.DataFrame(artist_frec)
df_artistTop.head(top)
```

```python
fig_top = plt.figure(figsize=(8,8))
plt.scatter(artist_frec['Artist'][:top],artist_frec['Frecuency'][:top])
plt.xticks(rotation=45);
plt.title('Artistas mas nominados en The Hot 100')
fig_top.show()

fig_top.savefig('Artistas más premiados.png')
```

**Song-Frecuency │ Artist**

```python
artist = artist_frec['Artist'][0]
song_frec = df[df['artist']==artist]['song'].value_counts()
song_frec = {'Canciones':list(song_frec.index),
              'Veces':list(song_frec.values)}

df_artist = pd.DataFrame(song_frec)

print(artist)
df_artist.head(10)
```
