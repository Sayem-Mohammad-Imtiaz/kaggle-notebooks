#### 2) (5 points) Compute the central tendency of any variable (mean, median, mode, range, and midrange).

```python
import statistics

#Nan's dropped for compute reals values.
Fare=titanic.Fare
Age_drop=titanic.Age.dropna(axis="rows")
Fare_drop=Fare.dropna(axis="rows")


print("Mean of Fare     :", np.mean(Fare_drop))
print("Median of Fare   :", np.median(Fare_drop))
print("Mode of Fare     :", statistics.mode(Fare_drop))
 
minval=min(Fare_drop)
maxval=max(Fare_drop)
print("Range of Fare    :", maxval-minval)
print("Mid-Range of Fare:", (maxval+minval)/2)
```

#### 3) (5 points) Compute the dispersion of data (standard deviation and variance).

```python
print("Standart Deviation of Fare:", statistics.stdev(Fare_drop) )
print("Variance of Fare          :", statistics.variance(Fare_drop) )
```

#### 4) (5 points) Compute 5-number summary of a distribution.

```python
quartiles = np.percentile(Fare_drop, [25, 50, 75])
print('Min   : %.3f' % minval)
print('Q1    : %.3f' % quartiles[0])
print('Median: %.3f' % quartiles[1])
print('Q3    : %.3f' % quartiles[2])
print('Max   : %.3f' % maxval)
```
