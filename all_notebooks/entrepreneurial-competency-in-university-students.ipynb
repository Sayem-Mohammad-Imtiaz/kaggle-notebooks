```python


import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.preprocessing import StandardScaler as SS
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import confusion_matrix, accuracy_score

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

```python
df = pd.read_csv('/kaggle/input/entrepreneurial-competency-in-university-students/data.csv')
```

## Data Preprocessing

```python
df.isna().sum()
```

```python
cols_miss = ['ReasonsForLack']
df = df.drop(cols_miss, axis = 1)
```

```python
df.isna().sum()
```

```python
df
```

```python
df.nunique()
```

```python
cols_le = ['EducationSector','KeyTraits']
cols_oh = ['Gender','City','Influenced','MentalDisorder']
```

```python
df['KeyTraits'].unique()
```

```python
LE = LabelEncoder()
df[cols_oh] = df[cols_oh].apply(LE.fit_transform)
```

```python
df['IndividualProject'] = LE.fit_transform(df['IndividualProject'])
```

```python
df[cols_le] = df[cols_le].astype(str)
df = pd.get_dummies(df , columns = cols_le)
```

```python
df
```

```python
df[cols_oh]
```

```python
df.shape
```

```python
df.keys()
```

```python
X = df.drop('y',axis = 1)
```

```python
y = df.y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

## Feature Scaling

```python
X_train.keys()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train[['Age','Perseverance','DesireToTakeInitiative','Competitiveness','SelfReliance','StrongNeedToAchieve','SelfConfidence','GoodPhysicalHealth']] = sc.fit_transform(X_train[['Age','Perseverance','DesireToTakeInitiative','Competitiveness','SelfReliance','StrongNeedToAchieve','SelfConfidence','GoodPhysicalHealth']])
X_test[['Age','Perseverance','DesireToTakeInitiative','Competitiveness','SelfReliance','StrongNeedToAchieve','SelfConfidence','GoodPhysicalHealth']] = sc.transform(X_test[['Age','Perseverance','DesireToTakeInitiative','Competitiveness','SelfReliance','StrongNeedToAchieve','SelfConfidence','GoodPhysicalHealth']])
```

## Classification Models

### Logistic Regression

```python
classifier1 = LogisticRegression()
classifier1.fit(X_train,y_train)
```

```python

y_pred1 = classifier1.predict(X_test)
cm = confusion_matrix(y_test, y_pred1)
print(cm)
accuracy_score(y_test, y_pred1)
```

### SVM

```python
classifier2 = SVC(kernel = 'poly', degree = 5)
classifier2.fit(X_train, y_train)
```

```python

y_pred2 = classifier2.predict(X_test)
cm = confusion_matrix(y_test, y_pred2)
print(cm)
accuracy_score(y_test, y_pred2)
```

```python
classifier3 = SVC(kernel = 'rbf')
classifier3.fit(X_train, y_train)
```

```python

y_pred3 = classifier3.predict(X_test)
cm = confusion_matrix(y_test, y_pred3)
print(cm)
accuracy_score(y_test, y_pred3)
```

### K-Nearest

```python
classifier4 = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier4.fit(X_train, y_train)
```

```python
y_pred4 = classifier4.predict(X_test)
cm = confusion_matrix(y_test, y_pred4)
print(cm)
accuracy_score(y_test, y_pred4)
```

### Naive Bayes

```python
classifier5 = GaussianNB()
classifier5.fit(X_train, y_train)
```

```python
y_pred5 = classifier5.predict(X_test)
cm = confusion_matrix(y_test, y_pred5)
print(cm)
accuracy_score(y_test, y_pred5)
```

### Decision Tree

```python
classifier6 = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier6.fit(X_train, y_train)
```

```python
y_pred6 = classifier6.predict(X_test)
cm = confusion_matrix(y_test, y_pred6)
print(cm)
accuracy_score(y_test, y_pred6)
```

### Random Forest

```python
classifier7 = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier7.fit(X_train, y_train)
```

```python
y_pred7 = classifier7.predict(X_test)
cm = confusion_matrix(y_test, y_pred7)
print(cm)
accuracy_score(y_test, y_pred7)
```
