```python
color=["#f94144","#f3722c","#f8961e","#f9c74f","#90be6d","#43aa8b","#577590"]
```

```python
#Ladder score
def plot_top_bottom(df_obj,column_name,col_desc,count):
    fig= plt.figure(figsize=(15,8))
    g=gs.GridSpec(ncols=1, nrows=2, figure=fig)
    plt.suptitle("Top "+str(count)+" and Bottom "+str(count)+" countries in "+col_desc+" 2021", family='Ariel', weight='bold', size=20)
    ax1=plt.subplot(g[0,0])

    top= df_obj.head(count)
    bot= df_obj.tail(count)
    ax1=sns.barplot(data=top, x=top[column_name],y=top['Country name'], color=color[4])
    #ax1.set_xlabel('')
    ax1.xaxis.set_visible(False)
    ax1.annotate("Top "+str(count) +" countries in ",xy=(8,2), family='Ariel', weight='bold', size=12)
    ax2=plt.subplot(g[1,0], sharex=ax1)
    ax2=sns.barplot(data=bot, x=bot[column_name],y=bot['Country name'], color=color[0])
    ax2.annotate("Bottom "+str(count)+" countries in "+col_desc+ " 2021",xy=(8,2), family='Serif', weight='bold', size=12)
    for s in ['left','right','top','bottom']:
        ax1.spines[s].set_visible(False)
        ax2.spines[s].set_visible(False)
```

# Top 10 and bottom 10 countries in Africa

```python
plot_top_bottom(dict_of_df['df_0'],'Ladder score','Happiness Index',10)
```
