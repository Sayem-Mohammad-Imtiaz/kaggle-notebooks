```python
#!pip install mlxtend --user

import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
import matplotlib.pyplot as plt




```

```python
data_path = '/kaggle/input/random-shopping-cart/dataset_group.csv'

```

```python
data = pd.read_csv(data_path, names=['Date','Count','Items'])


```

```python
transactionData = pd.crosstab(data['Date'],data['Items'])
```

```python
transactionData.sum().\
    to_frame('Frequency').\
    sort_values('Frequency',ascending=False)[:5].\
    plot(kind='bar',
         figsize=(12,8),
         title="Frequent Items")

plt.show()
```

```python
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
trans_sets = transactionData.applymap(encode_units)
trans_sets
```

```python
frequent_itemsets = apriori(trans_sets, 
                            min_support = 0.2, 
                            use_colnames = True)

frequent_itemsets
```

```python
frequent_itemsets['len'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
```

```python
rules = association_rules(frequent_itemsets, 
                          metric = "lift", 
                          min_threshold = 1)
rules.head()
```

```python
rules['len'] = rules.antecedents.apply(lambda x: len(x))
rules
```

```python
rules = rules[(rules.len >=2) & (rules.confidence > 0.75) & (rules.lift > 1.2)]
```

```python
rules = rules.sort_values(['support','confidence','lift'], ascending=[True,False,False])
rules
```
