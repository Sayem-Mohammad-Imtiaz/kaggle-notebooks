Logistic Regression

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../input/abstract-data-set-for-credit-card-fraud-detection/creditcardcsvpresent.csv')
data.dropna(axis=1,inplace=True) # Drop Transaction date
data
```

### Y/N -> 1/0

```python
data.loc[data['isFradulent']=='N','isFradulent'] = 0
data.loc[data['isFradulent']=='Y','isFradulent'] = 1

data.loc[data['Is declined']=='N','Is declined'] = 0
data.loc[data['Is declined']=='Y','Is declined'] = 1

data.loc[data['isForeignTransaction']=='N','isForeignTransaction'] = 0
data.loc[data['isForeignTransaction']=='Y','isForeignTransaction'] = 1

data.loc[data['isHighRiskCountry']=='N','isHighRiskCountry'] = 0
data.loc[data['isHighRiskCountry']=='Y','isHighRiskCountry'] = 1
```

```python
data.head()
```

```python
x = data.drop(['Merchant_id','isFradulent'],axis=1)
y = data['isFradulent'].astype(dtype=int)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_cols = ['Average Amount/transaction/day', 'Transaction_amount', 'Total Number of declines/day', 'Daily_chargeback_avg_amt', '6_month_avg_chbk_amt', '6-month_chbk_freq']
x[num_cols] = scaler.fit_transform(x[num_cols])
x.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 101)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)

preds = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, plot_confusion_matrix

model_acc = round(accuracy_score(y_test, preds) * 100, 3)

print(f'Logistic Regression Accuracy : {model_acc}%')
plot_confusion_matrix(model, x_test, y_test, cmap=plt.cm.Blues)
```
