Cases registered via states

```python
plt.figure(figsize=(10,10));
rape.groupby(["State/UT","2014 - Cases registered","2015 - Cases registered","2016 - Cases registered"]).sum().plot(kind="bar", width = 0.5, figsize=(10,10), title=" Cases Registered Against RAPE");
plt.plot();
plt.show()
```

```python
rape14 = rape[["State/UT","2014 - Cases registered","2014 - Total rape Cases"]]
print(rape14)
```

```python
sns.set()
plt.figure(figsize=(15,15))
sns.barplot(data = rape14)
plt.title("2014 rape cases",size=20)
plt.xlabel("Numbers")
plt.ylabel("Year")
plt.show()
```

2014 rape percentage

```python
rape["rape_2014_perc"]=(rape["2014 - Total rape Cases"]/rape["2014 - Cases registered"])*100
rape["rape_2014_perc"]= rape["rape_2014_perc"].map(lambda x: round(x,2))
```

```python
rape.rape_2014_perc
```

```python
# Same add for 2015 & 2016
rape["rape_2015_perc"] = (rape["2015 - Total rape Cases"]/rape["2015 - Cases registered"])*100
rape["rape_2015_perc"] = rape["rape_2015_perc"].map(lambda x: round(x, 2))
rape["rape_2016_perc"] = (rape["2016 - Total rape Cases"]/rape["2016 - Cases registered"])*100
rape["rape_2016_perc"] = rape["rape_2016_perc"].map(lambda x: round(x, 2))
```

```python
sns.lineplot(data=rape.rape_2014_perc)
plt.show()
```

```python
sns.lineplot(data=rape.rape_2015_perc)
plt.show()
```

```python
sns.lineplot(data=rape.rape_2016_perc)
plt.show()
```

**State wise Rape percentage**

```python
state_rape =rape.groupby('State/UT')["rape_2014_perc", "rape_2015_perc", "rape_2016_perc"].sum().plot.bar(stacked = True, figsize=(15,15), title="State vs Rape Percentage");
state_rape.set_xlabel("States");
state_rape.set_ylabel("%");
```
