```python
import numpy as np
import pandas as pd
import plotly.graph_objs as go
import plotly.express as px
```

```python
df = pd.read_csv("/kaggle/input/latest-worldwide-vaccine-data/Worldwide Vaccine Data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df[df.isnull().values==True]
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df = df.rename(columns = {'Doses administered per 100 people':'administered_per_100_people'})
df = df.rename(columns = {'Total doses administered':'total_administered'})
df = df.rename(columns = {'% of population vaccinated':'vaccinated'})
df = df.rename(columns = {'% of population fully vaccinated':'fully_vaccinated'})
```

```python
df.replace(',','',regex = True, inplace = True)
```

```python
df1 = df.sort_values(by = ['vaccinated'],ascending = False).reset_index().head(10)

fig = px.bar(df1, x = 'Country', y = 'vaccinated', 
             color = 'vaccinated', color_continuous_scale = 'blues',
             labels= {"vaccinated":"Vaccinated(%)"}
            )

fig.update_layout(title = 'Top 10 Countries (% of population vaccinated)',
                  title_x = 0.5,
                  title_font = dict(size = 22, family = 'Balto',color = 'Navy'),
                  yaxis = dict(title = 'Vaccinate(%)'))

fig.show()
```

```python
df1 = df.sort_values(by = ['fully_vaccinated'],ascending = False).reset_index().head(10)

fig = px.bar(df1, x = 'Country', y = 'fully_vaccinated', 
             color = 'fully_vaccinated', color_continuous_scale = 'redor',
             labels= {"fully_vaccinated":"Fully vaccinated(%)"})

fig.update_layout(title = 'Top 10 Countries (% of population fully vaccinated)',
                  title_x = 0.5,
                  title_font = dict(size = 22, family = 'Balto',color = 'FireBrick'),
                  yaxis = dict(title = 'Fully Vaccinate(%)'))

fig.show()
```

```python
fig = go.Figure(data = go.Choropleth(locations = df['Country'],
                                     z = df['administered_per_100_people'],
                                     locationmode = 'country names',
                                     colorscale = 'twilight',
                                     autocolorscale = False,
                                     reversescale = True,
                                     marker_line_color = 'LightGray',
                                     marker_line_width = 0.3,
                                     colorbar_title = 'Doses Administered/per 100 people'))

fig.update_traces(colorbar_title_font = dict(family = 'Droid Sans',
                                             size = 12,
                                             color = 'DarkBlue'))
fig.update_layout(
    title_text = 'Countries in the world (Doses Administered per 100 people)',
    title_font = dict(size = 22, family = 'Balto',color = 'DarkBlue'),
    title_x = 0.5,
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection_type= 'equirectangular'))
fig.show()  
```

```python
fig = go.Figure(data = go.Choropleth(locations = df['Country'],
                                     z = df['total_administered'],
                                     locationmode = 'country names',
                                     colorscale = 'cividis',
                                     autocolorscale = False,
                                     reversescale = True,
                                     marker_line_color = 'LightGray',
                                     marker_line_width = 0.3,
                                     colorbar_title = 'Total doses administered'))

fig.update_traces(colorbar_title_font = dict(family = 'Droid Sans',
                                             size = 12,
                                             color = 'DarkBlue'))
fig.update_layout(
    title_text = 'Countries in the world (Total doses administered)',
    title_font = dict(size = 22, family = 'Balto',color = 'DarkBlue'),
    title_x = 0.5,
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection_type= 'equirectangular'))
fig.show()  
```

```python
df1 = df.sort_values(by = ['vaccinated'],ascending = False).reset_index()

fig = px.choropleth(df1,
                    locations = 'Country',
                    locationmode = 'country names',
                    color = 'vaccinated',
                    color_continuous_scale = 'plasma',
                    hover_name = 'Country',
                    animation_frame = 'Country',
                    labels= {"vaccinated":"Vaccinated(%)"})

fig.update_layout(title = 'Countries in the world (% of population vaccinated)',
                  title_x = 0.5,
                  title_font = dict(size = 22, family = 'Balto',color = 'DarkSlateBlue'),
                  geo = dict(showframe = False,
                             showcoastlines = False,
                             projection_type = 'natural earth'))
fig.show()
```

```python
df1 = df.sort_values(by = ['fully_vaccinated'],ascending = False).reset_index()

fig = px.choropleth(df1,
                    locations = 'Country',
                    locationmode = 'country names',
                    color = 'fully_vaccinated',
                    color_continuous_scale = 'reds',
                    hover_name = 'Country',
                    animation_frame = 'Country',
                    labels= {"fully_vaccinated":"Fully vaccinated(%)"})

fig.update_layout(title = 'Countries in the world (% of population fully vaccinated)',
                  title_x = 0.5,
                  title_font = dict(size = 22, family = 'Balto',color = 'DarkRed'),
                  geo = dict(showframe = False,
                             showcoastlines = False,
                             projection_type = 'natural earth'))
fig.show()
```
