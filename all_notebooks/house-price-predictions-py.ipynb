```python

from sklearn.base import TransformerMixin

for i in features1.columns:
    if features1[i].dtype==('object'):
     features1[i].fillna(features1[i].mode()[0],inplace=True)
    else:
     features1[i].fillna(features1[i].mean(),inplace=True)
    
    
    
   
```

```python
print(features1.info())
print(features1)

features2=features1

print(features1)
features1=features1.iloc[:,:-1]
print(features1)
```

```python
for j in testfeatures.columns:
    if testfeatures[j].dtype==('object'):
     testfeatures[j].fillna(testfeatures[j].mode()[0],inplace=True)
    else:
     testfeatures[j].fillna(testfeatures[j].mean(),inplace=True)
```

```python
print(testfeatures)
```

```python
print(features1.shape)
```

```python
categorical_feature_mask = features1.dtypes==object
categorical_cols = features1.columns[categorical_feature_mask].tolist()
# import labelencoder
from sklearn.preprocessing import LabelEncoder
# instantiate labelencoder object
le = LabelEncoder()
features1[categorical_cols] = features1[categorical_cols].apply(lambda col: le.fit_transform(col))
features1[categorical_cols].head(10)
print(features1)
```

```python
categorical_feature_mask1 = testfeatures.dtypes==object
categorical_cols1 = testfeatures.columns[categorical_feature_mask1].tolist()
# import labelencoder
from sklearn.preprocessing import LabelEncoder
# instantiate labelencoder object
le1 = LabelEncoder()
testfeatures[categorical_cols1] = testfeatures[categorical_cols1].apply(lambda col: le1.fit_transform(col))
testfeatures[categorical_cols1].head(10)
print(testfeatures)
```

```python
# Import the model we are using
from sklearn.ensemble import RandomForestRegressor
# Instantiate model with 1000 decision trees
rf = RandomForestRegressor(n_estimators = 1000, random_state = 42)
# Train the model on training data
rf.fit(features1, traintarget)
print(rf.score(features1,traintarget))
```

```python
pred2=rf.predict(testfeatures)
predrf=pd.DataFrame(pred2)
print(predrf)

                   
                 
```
