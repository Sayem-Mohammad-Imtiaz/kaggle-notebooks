```python
res = pd.read_csv("/kaggle/input/3k-twitter-giveaway-data/DatasetGiveawaysHandmade.csv")
print(res)
res= res[~res.text.str.startswith('RT')]
print(res)
```

```python
train, test = train_test_split(res, random_state=42, test_size=0.2, shuffle=True)
X_train = train.text
X_test = test.text

print(X_train.shape)
print(X_test.shape)
```

```python
# Define a pipeline combining a text feature extractor with multi lable classifier
NB_pipeline = Pipeline([
                ('tfidf', TfidfVectorizer(stop_words=stop_words)),
                ('clf', OneVsRestClassifier(MultinomialNB(
                    fit_prior=True, class_prior=None))),
            ])

NB_pipeline.fit(X_train, train.giveaway)

prediction = NB_pipeline.predict(X_test)
print("Accuracy :",accuracy_score(test.giveaway, prediction))
```

```python
SVC_pipeline = Pipeline([
                ('tfidf', TfidfVectorizer(stop_words=stop_words)),
                ('clf', OneVsRestClassifier(LinearSVC(), n_jobs=1)),
            ])
SVC_pipeline.fit(X_train, train.giveaway)
# compute the testing accuracy
prediction = SVC_pipeline.predict(X_test)
print("Accuracy :",accuracy_score(test.giveaway, prediction))
```
