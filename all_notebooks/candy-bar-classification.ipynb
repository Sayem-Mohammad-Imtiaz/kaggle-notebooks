# Predicting if a given candy is a bar or not

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

**Splitting data**

```python
df1=df.copy()
df1.drop("competitorname", axis=1, inplace=True)

X = df1.drop(["bar", "hard", "sugarpercent", "winpercent"], axis=1)
y =df1["bar"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 101)
```

# Choosing K value

```python
error = []

for i in range(1,50):
    knn = KNeighborsClassifier(n_neighbors = i )
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    error.append(np.mean(pred != y_test))

plt.figure(figsize=(13,8))
plt.xlabel("K")
plt.ylabel("Error")
sns.lineplot(range(1,50), error, marker = "*" , markersize = 15, color="black")
```

```python
models = {

    "                   Logistic Regression": LogisticRegression(),
    "        K-Nearest Neighbors Classifier": KNeighborsClassifier(n_neighbors=13),
    "                        SVM Classifier": SVC(),
}
```

# Training

```python
for name, model in models.items():
    model.fit(X_train, y_train)
    print(name + " trained.")
```

# Testing

```python
for name, model in models.items():
    print(name + " Accuracy: {:.5f}".format(model.score(X_test, y_test)))
                     
```
