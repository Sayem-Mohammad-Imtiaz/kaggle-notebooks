```python
#Training Linear Regression
```

```python
X = housing_df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
y = housing_df['Price']
```

```python
#Train Test Spliting
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
#Model Creation and training
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr_model = LinearRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
#Evaluation of Trained Model
```

```python
print(lr_model.intercept_)
```

```python
#'''price = c1*'Avg. Area Income'+ c2* 'Avg. Area House Age'+ c3*'Avg. Area Number of Rooms
 #      +c4*'Avg. Area Number of Bedrooms' +c5* 'Area Population''''
```

```python
coef_hosing = pd.DataFrame(lr_model.coef_, X.columns, columns=['Coefficient'])
```

```python
coef_hosing
```

```python
# If all other features are fixed, then 1 unit change in Avg. Area Income Price would increase 21.60 Rupees
```

```python
# Making Prediction by making use of Trained Model
```

```python
predictions = lr_model.predict(X_test)
```

```python
plt.scatter(y_test, predictions)
```

```python
sns.distplot((y_test-predictions))
```

```python
from sklearn import metrics
```

```python
mae = metrics.mean_absolute_error(y_test, predictions)
mse = metrics.mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)
```

```python
print('Mean Absolute Error={}\nMean Square Error={}\nRoot Mean Square Error={}'.format(mae,mse,rmse))
```
