# Mission1: Global animation of COVID vaccination cases per country 

**Code and result:**

```python
import numpy as np
import pandas as pd
import plotly.express as px
import plotly

df=pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
df.index=pd.to_datetime(df['date'])
df=df.sort_index()
#print(df.columns)
fig=px.choropleth(df,locations='country',locationmode='country names',color='daily_vaccinations',hover_name="country",animation_frame="date",color_continuous_scale="Viridis",range_color=(0.0,800000.0))
fig.update_layout(title_text="Daily vaccinated population in each country",title_x=0.5)
fig.show()

#plotly.offline.plot(fig,filename="vaccinations.html")
```

**Description:**

I did an analysis of the daily vaccinated population in each country. I chose this column because it will give a clear picture of which countries are promoting vaccination at the highest pace.

For the color bar, I chose viridis instead of rainbow. The reason is that viridis consists of colors in the same color system, hence we can judge the values simply base on the color depth.

The results of the analysis are as shown. Observing the recent dates, we found that China, the US, India, and Brazil have the most fastest vaccination speed. (brightest color) The number of daily vaccine injections in these countries can exceed millions, but when the upper limit of the range was set at about millions, the numbers in the other countries were compared and stayed in a deep color. Therefore, I set the upper limit to 800k in order to show the differences between countries in an understandable way.

# Mission2: Hourly animation of global earthquake locations 

**Code and result:**

```python
df=pd.read_csv("../input/earthquakedata20100112/query.csv")
df.index=pd.to_datetime(df['time'])
df['time']=df.index.strftime('%Y-%m-%d %H:00:00')
df=df.sort_index()
fig2=px.scatter_geo(df,lat='latitude',lon='longitude',color='mag',animation_frame='time',color_continuous_scale='rainbow',range_color=(1.0,9.0))
fig2.update_layout(title_text="Earthquake happened on 12nd January 2010",title_x=0.5)
fig2.show()
```

**Description**

The analysis was conducted on January 12, 2010, the day when Haiti earthquake occurred. Since it was an unprecedented catastrophe that occurred in Haiti, one of the poorest countries in the world, it had left a lasting impression on me since I was a child.

By observing the timeline, it can be confirmed that the earthquake of magnitude 7.0 occurred in Haiti at approximately 21:00 UTC time. Aftershocks of aorund magnitude 5.0 was seen in the same location an/two hour later.
