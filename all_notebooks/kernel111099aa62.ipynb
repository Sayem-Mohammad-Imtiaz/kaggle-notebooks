```python
labels = df['class']
df.drop('class' , inplace = True , axis = 1 )
print(labels.unique())
```

```python
listt = df.columns
dict = []
for i in listt :
    dict.append(( len(df[i].unique()) / df.shape[0] , i ))
```

```python
dict = sorted(dict)
for i , j in   dict:
    print( i , j )
```

```python
# drop first 2 columns 
df.drop(['objid' , 'rerun'] , inplace = True , axis = 1 )
```

```python
dict2 = []
for i in df.columns :
    dict2.append( ( df[i].isnull().sum() , i ) )
dict2 = sorted(dict2 , reverse = False)
for i , j in dict2 :
    print( i , j )
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
df = ss.fit_transform(df)
```

```python
df1 = pd.DataFrame(df)
```

```python
le = preprocessing.LabelEncoder()
labels = le.fit_transform(labels)
print(labels)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df1, labels, test_size=0.33)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC(kernel='rbf')
clf.fit(X_train, y_train)
```

```python
print(clf.score(X_test , y_test))
from sklearn.metrics import f1_score

y_pred = clf.predict(X_test)
print(f1_score(y_test, y_pred, average='macro'))
print(f1_score(y_test, y_pred, average='micro'))
print(f1_score(y_test, y_pred, average='weighted'))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
clf.fit(X_train,y_train)
print(clf.score(X_test , y_test))
from sklearn.metrics import f1_score

y_pred = clf.predict(X_test)
print(f1_score(y_test, y_pred, average='macro'))
print(f1_score(y_test, y_pred, average='micro'))
print(f1_score(y_test, y_pred, average='weighted'))
```
