```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
df = pd.read_csv('../input/heights-and-weights/data.csv', header='infer')
len(df)
```

(Or how to load a simple model and dump its parameters using joblib.)

```python
df.head()
```

lets convert height to cm

```python
df['HeightCM'] = df['Height'] * 100
```

```python
df.head()
```

```python
X = df.loc[:,df.columns =='HeightCM']
y = df['Weight']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

lr = LinearRegression().fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.score(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

now using joblib to dump the model

```python
import joblib 
joblib.dump(lr, 'linear_regression_model.pkl')
```
