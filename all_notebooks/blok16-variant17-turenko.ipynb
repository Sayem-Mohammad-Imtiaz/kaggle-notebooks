```python
# Задание 1.

import pandas as pd
import numpy as np
import statistics as st
import matplotlib.pyplot as plt
weatheraus = pd.read_csv('../input/weatheraus/weatherAUS.csv')
weatheraus
```

```python
# Задание 2. Статистические показатели

print('mediana:', st.median(weatheraus['WindSpeed9am'])) # медиана
print('moda:', st.mode(weatheraus['WindSpeed9am'])) # мода
print('mean:', st.mean(weatheraus['WindSpeed9am'])) # средняя
print('min:', np.min(weatheraus['WindSpeed9am'])) # минимум
print('max:', np.max(weatheraus['WindSpeed9am'])) # максимум
```

```python
# Задание 3. Разница между максимумом и минимумом

weatheraus['raznica'] = weatheraus.max(axis=1) - weatheraus.min(axis=1)
print(weatheraus)
```

```python
# Задание 4. Разбиваем данные на несколько блоков

blok1 = weatheraus[(weatheraus['WindSpeed9am'])<=10]
print('скорость ветра меньше 10:',len(blok1.index))

blok2 = weatheraus[(weatheraus['WindSpeed9am'])>10 &(weatheraus['WindSpeed9am']<=15)]
print( 'скорость ветра от 10 до 15:',len(blok2.index))

blok3 = weatheraus[(weatheraus['WindSpeed9am'])>15]
print('скорость ветра больше 15:',len(blok3.index))
```

```python
# Задание 5. Круговая диаграмма для разбиения

labels = ['<10','10-15','>15']
values = [len(blok1.index),len(blok2.index),len(blok3.index)]
colors = ['yellow','green','blue']

fig1, ax1 = plt.subplots()
ax1.pie(values, labels=labels)
plt.show()
```

```python
# Задание 6. Для каждого блока строим график

# blok1
plt.plot(blok1['WindSpeed9am'], blok1['Humidity9am'])
plt.show()
plt.hist(blok1['WindSpeed9am'], color='green',alpha=0.7,rwidth=0.85)
plt.show()

# blok2
plt.plot(blok2['WindSpeed9am'], blok2['Humidity9am'])
plt.show()
plt.hist(blok2['WindSpeed9am'], color='green',alpha=0.7,rwidth=0.85)
plt.show()

# blok3
plt.plot(blok3['WindSpeed9am'], blok3['Humidity9am'])
plt.show()
plt.hist(blok3['WindSpeed9am'], color='green',alpha=0.7,rwidth=0.85)
plt.show()
```

```python
# Задание 7. Группировка данных по некоторому признаку

weatheraus['Date'] = pd.to_datetime(weatheraus['Date'])
weatheraus['year'] = weatheraus['Date'].dt.year
weatheraus_group=dict(tuple(weatheraus.groupby('year')))

group1 = weatheraus_group[2008].sort_values(by=['Date', 'MinTemp'], ascending=[True,False])
print(group1)

group2 = weatheraus_group[2009].sort_values(by=['Date', 'MinTemp'], ascending=[True,False])
print(group2)

group3 = weatheraus_group[2010].sort_values(by=['Date', 'MinTemp'], ascending=[True,False])
print(group3)
```
