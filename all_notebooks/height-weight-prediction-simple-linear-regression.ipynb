```python
# importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# loading dataset
df = pd.read_csv('../input/heights-and-weights/data.csv')
df
```

```python
df.describe()
```

```python
df.info()
```

```python
# filling null values
df.isnull().mean()
```

```python
# getting independent variable
x = df.drop(['Weight'],axis = 1)
```

```python
x
```

```python
# getting dependent variable or target variable
y = df['Weight']
```

```python
y
```

```python
# plot height vs weight
plt.scatter(x,y)
plt.xlabel('height')
plt.ylabel('weight')
plt.title('height vs weight')
```

```python
# box plot for the distribution of the data
plt.boxplot(x)
plt.title('height')
plt.xlabel('height')
plt.ylabel('distribution')
```

```python
# from above we are not able to get any outliners so we donot need to do any kind of outliner handling.
```

```python
# train test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = .3)
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
model = LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
y_test
```

```python
model.intercept_
```

```python
model.coef_
```

```python
# equation of y_pred = mx+c
y_pred = (model.coef_*x_test)+model.intercept_
y_pred
```

```python
plt.plot(x_test,y_pred, color = 'red')
plt.scatter(x,y)
plt.xlabel('height')
plt.ylabel('weight')
plt.title('simple regression model ')
```

```python
# checking how accurately data is being predicted or model is working
from sklearn.metrics import mean_squared_error,r2_score
```

```python
# mean square error: lower the value more accurate is the prediction
mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
# this is equal to model score giving the accuracy of model(it lies between 0-1). Higher the value greater the accuracy.
r2_score(y_test,model.predict(x_test))
```

```python
model.score(x_test,y_test)
```

```python
# from the above result the model comes out be 98.46% accurate which is quite good result.
```
