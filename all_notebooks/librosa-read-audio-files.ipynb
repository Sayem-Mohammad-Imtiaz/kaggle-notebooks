```python
data_dir = '../input/set_a'
audio_files = glob(data_dir+'/*.wav')
```

```python
audio, sfreq = lr.load(audio_files[0])
time = np.arange(0,len(audio))/sfreq
```

```python
fig, ax = plt.subplots()
ax.plot(time,audio)
```

```python
for file in range(0, len(audio_files),1):
    audio, sfreq = lr.load(audio_files[file])
    fig,ax = plt.subplots()
    time = np.arange(0,len(audio))/sfreq
    ax.plot(time,audio)
    plt.show()
```
