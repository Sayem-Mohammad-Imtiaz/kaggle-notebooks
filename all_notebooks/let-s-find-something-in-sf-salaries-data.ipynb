EACH STATUS JOB MEDIAN

```python
tempFT=temp[temp['Status']=='FT']
tempPT=temp[temp['Status']=='PT']
temp0=temp[~temp['Status'].isin(['PT','FT'])]
```

```python
print('PT',tempPT['TotalPay'].median(),',','FT',tempFT['TotalPay'].median(),',','unknown',temp0['TotalPay'].median())
```
