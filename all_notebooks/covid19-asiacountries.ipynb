```python
#create a new column to calculate the percentage of infected people with total population
covid_data["Percentage of infected"]=covid_data["Total Cases"]*100/covid_data["Population"]
```

```python
#sort the top 10 of highest infected per population
covid_data.sort_values("Percentage of infected", ascending=False).head(10)
```

```python
covid_data.sort_values("Total Cases per 1M population", ascending=False).head(10)
```

I believe the result of percentage of infected will return the same result as the total cases per 1M population when sorted by highest to lowest.

```python
#find which countries suffered the most death per 1M population
covid_data.sort_values("Deaths per 1M population", ascending=False).head(10)
```

```python
#calculate the fatality rate of the disease in each countries
covid_data["Fatality rate in percentage"]=covid_data["Total Deaths"]*100/covid_data["Total Cases"]
covid_data
```

```python
#top 15 fatality rate
covid_data.sort_values("Fatality rate in percentage", ascending=False).head(15)
```

```python
#take top 5 for the data visualisation
top_five=covid_data.sort_values("Total Cases per 1M population", ascending=False).head(5)
top_five
```

Plot the graph for the 5 highest percentage of infected population countries

```python
sns.relplot(x="Deaths per 1M population", y="Total Cases per 1M population",size="Active Cases", hue="Country", data=top_five)
```
