```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def calculateMeanandFill(df):
    for column in df.columns:
        if column == "type":
            continue
        df[column].fillna(df[column].mean(), inplace = True)
    else:
        df.dropna(axis = 0, inplace = True)
        return True
```

```python
train = pd.read_csv("/kaggle/input/wine-quality/winequalityN.csv")
calculateMeanandFill(train)
train['type'] = train['type'].replace(['white', 'red'], [0, 1])
y = train.pop("quality")
```

```python
from sklearn.tree import DecisionTreeRegressor

wine_model = DecisionTreeRegressor(random_state=1)

# Fit Model
wine_model.fit(train, y)
```

```python
# Predictions
import random
index = random.randint(0,6498)

# Values
print("Values:\n")
print(train.loc[index])
print("\nExpected Result:")
print(y.loc[index])
print("\nPrediction:")
print(wine_model.predict([train.loc[index]])[0])
```
