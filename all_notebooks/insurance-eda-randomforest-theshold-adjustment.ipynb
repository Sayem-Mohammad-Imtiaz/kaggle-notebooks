## Data Cleaning in preparation for model

turning those who 'yes/no' to 1/0 - GraduateOrNot, FrequentFlyer, EverTravelledAbroad

turning employment type into 1/0 too - Governemnt Sector -> 1, Private Sector -> 0

```python
#function for turning yes or no

def clean_string(x):
    if x == 'Yes':
        return 1
    else:
        return 0
```

```python
#function for profession
def employment(x):
    if x == 'Government Sector':
        return 1
    else:
        return 0
```

```python
df
```

```python
df_clean = df
df_clean['GraduateOrNot'] = df_clean['GraduateOrNot'].apply(clean_string)
df_clean['FrequentFlyer'] = df_clean['FrequentFlyer'].apply(clean_string)
df_clean['EverTravelledAbroad'] = df_clean['EverTravelledAbroad'].apply(clean_string)
df_clean['Employment Type'] = df_clean['Employment Type'].apply(employment)
```

```python
df_clean
```

## Feature Engineering

Using KNN first to give each record a "Customer Score".  *I read this idea from Practical Stats for Data Scientist, in which it talked about using KNN predict_prob as a way to assign a pseudo-probability*

I would need to scale of age, annual income and family members before I add the KNN probability

```python
df_knn_prep = df_clean.copy() #create a copy
```

```python
#import required library

from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
to_scale = ['AnnualIncome','Age','FamilyMembers']
scaler = MinMaxScaler()
```

```python
df_knn_prep[to_scale] = scaler.fit_transform(df_knn_prep[to_scale])
```

```python
df_knn_prep
#cleaned up ready for KNN
```

```python
knn_model = KNeighborsClassifier(n_neighbors = 20)

#define X
knn_X = df_knn_prep[['Age','Employment Type','GraduateOrNot','AnnualIncome','FamilyMembers','ChronicDiseases','FrequentFlyer','EverTravelledAbroad']]
knn_y = df_knn_prep['TravelInsurance']


knn_model.fit(knn_X, knn_y) #using the whole dataset since this is not our final model, but a pseudo-probability "Customer Score"
df_clean['CustomerScore'] = knn_model.predict_proba(knn_X)[:,1]
```

```python
#customer score added, which will be included in the model
X = df_clean[['Age','Employment Type','GraduateOrNot','AnnualIncome','FamilyMembers','ChronicDiseases','FrequentFlyer','EverTravelledAbroad','CustomerScore']]
y = df_clean['TravelInsurance']
df_clean
```

## RandomForestClassifier and GridSearchCV

In my opinion, recall should be the important scoring criteria.  It is paramount that we capture those likely to buy Travel Insurance as best as we could since that would be lost revenue otherwise.

```python
#import model and hyperparameter tuning

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.1,random_state=42)
#split to validate best parameters later
```

```python
rfc = RandomForestClassifier()

param = {'n_estimators':[50], 'criterion':['gini'], 'max_depth':[25,20,15,10], 'min_samples_split':[2,5,10,15,20], }

cv = GridSearchCV(estimator=rfc,param_grid=param, scoring = 'recall',n_jobs=-1,cv=5,verbose=5)
```

```python
cv.fit(X_train, y_train)
```

```python
cv.best_params_
```

```python
cv.best_score_ #recall
```

```python
y_pred = cv.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

## Trying different probability threshold

The above is our baseline results with the best parameteres, maybe we can use predict_proba and some function to output a class prediction according to some threshold to improve the recall

```python
rfc = RandomForestClassifier(criterion='gini',
 max_depth = 25,
 min_samples_split = 2,
 n_estimators = 50)
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred_proba = rfc.predict_proba(X_test)[:,1]
```

```python
#function to get class based on probability
def get_y_pred(p, proba_pred):
    y_pred = []
    for i in range(0,len(proba_pred)):
        if proba_pred[i] > p:
            y_pred.append(1)
        else:
            y_pred.append(0)
    return pd.Series(y_pred)
```

```python
prob = [0.1,0.15,0.2,0.25,0.3,0.35,0.4]

for p in prob:
    y_pred = get_y_pred(p,y_pred_proba)
    print(f'At P = {p}, Recall score = {recall_score(y_test,y_pred)}')
    print(f'At P = {p}, Accuracy score = {accuracy_score(y_test,y_pred)}')
    print(confusion_matrix(y_test,y_pred))
    print('\n')
```

Depending on the aggressiveness of the company's sales practice, the P can be adjusted
