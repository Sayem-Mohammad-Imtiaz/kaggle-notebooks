```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../input/greenhouse-gas-emissions/GHG-Emissions-by-sector.csv').set_index('Sub-sector')
data.head()
```

```python
data.shape
```

```python
data.describe()
```

```python
data.GlobalShare.sum()
```

# Data Distribution Plot

```python
fig, ax = plt.subplots(1,3, figsize=(15,8))
sns.boxplot(y=data.GlobalShare, ax=ax[0], linewidth=2.5, color='#7FFFD4')
sns.stripplot(y=data.GlobalShare, ax=ax[1], s=15, alpha=0.8, color='#191970')
sns.violinplot(y=data.GlobalShare, ax=ax[2], color='#00CED1')
plt.show()
```

```python
data = data.sort_values('GlobalShare', ascending=False)
data
```

```python
print(f'FINDINGS: Green House Gas Emission Maximum by {data.index[0]} ({data.GlobalShare[0]}%) & Minimum by {data.index[-1]} ({data.GlobalShare[-1]}%)')
```

# BAR PLOT

```python
# BAR PLOT
plt.figure(figsize=(12,10))
plt.title('Global Share of Green House Gas Emission (%)', fontsize=15)
plot = sns.barplot(data.GlobalShare, data.index, palette='viridis')
plt.bar_label(plot.containers[0],size=12)
plt.show()
```

# PIE CHART

```python
# PIE CHART
data.GlobalShare.plot.pie(autopct='%1.2f%%',figsize=(15,15), 
                          label='GHG Emission', shadow=True)
plt.show()
```

# SCATTER PLOT

```python
# SCATTER PLOT
plt.figure(figsize=(15,10))
sns.scatterplot(data.GlobalShare, data.index, s=100, color='#8B008B', alpha=0.8)
for line in range(0,len(data)):
     plt.text(data.GlobalShare[line]+0.1, data.index[line], data.GlobalShare[line])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title("Top Sectors in high GHG Emission", fontsize=15)
plot1 = sns.barplot(data.index[:7], data.GlobalShare[:7], palette='viridis')
plt.bar_label(plot1.containers[0],size=12)
plt.xticks(rotation='vertical')
plt.show()
```
