### In this work I analyze my Fitbit Charge data. The data spans over two years and contains the number of steps walked per day, the number of floors, minutes spent sitting, minutes of slow, moderate or intense activity as well as the number of calories.

```python
import pandas as pd 
import numpy as numpy 
import matplotlib.pyplot as plt
import datetime
import calendar 
  
def findDay(date): 
    born = datetime.datetime.strptime(date, '%d %m %Y').weekday() 
    return (calendar.day_name[born]) 
```

```python
fitbit = pd.read_csv('../input/one-year-of-fitbit-chargehr-data/One_Year_of_FitBitChargeHR_Data.csv') # fitbit data
fitbit.shape
```

```python
fitbit.head()
```

```python
# convert the data type and set Date as index
fitbit['Date'] = pd.to_datetime(fitbit['Date'])
fitbit = fitbit.set_index('Date')
```

### Activity on Business days VS Weekends 

```python
# create lists for weekends and business days
activity_wd = []
activity_bd = []

# reset index for obtaining the row Date
fitbit = fitbit.reset_index()
```

```python
# loop through the rows
counter = 0
for index, row in fitbit.iterrows(): 
    
    date = str(row['Date'])
    msi = row['Minutes_of_intense_activity']
    
    date = date.split(' ')
    
    date = date[0].replace('-', ',')
    
    year = date.split(',')[0]
    
    month = date.split(',')[1]
  
    day = date.split(',')[2]
    
    day_string = str(day)+' '+str(month)+' '+str(year)
    print(day_string)
 
    day_of_week = findDay(day_string)
    if day_of_week == 'Saturday' or day_of_week == 'Sunday': 
        activity_wd.append(msi)
    else: 
        activity_bd.append(msi)
```

```python
# Plot the time series of floors per day

plt.figure(figsize=(20,4))
plt.plot(activity_bd)
plt.show()

plt.figure(figsize=(20,4))
plt.plot(activity_wd)
plt.show()
```
