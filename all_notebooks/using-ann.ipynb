```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
import tensorflow as tf
import keras as keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
data = pd.read_csv("../input/predicting-churn-for-bank-customers/Churn_Modelling.csv")
data.head()
```

```python
x = data.iloc[:,3:-1]
y = data.iloc[:,-1]
x.head()
```

# Converting Categorical into numerical values
    
   1. Dummy variables
   2. Label Encoder
   
 For Geography, I am using Dummies replacement as there are 3 different values.
 For Gender, I will go with Label encoder(I can get only 0,1)

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
```

```python
le = LabelEncoder()
x['Gender'] = le.fit_transform(x['Gender'])
```

```python
x = pd.get_dummies(x,drop_first=True,columns=['Geography'])
x.head()
```

# Making data into Standard Distribution

As we can see the data is not in same range for all columns, for this we have to make standardization.

```python
x_train,x_test,y_train , y_test = train_test_split(x,y,test_size=.2,stratify=y,random_state=10)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x.shape
```

# ANN

```python
model = Sequential()
model.add(Dense(x.shape[1],activation='relu',input_dim=x.shape[1]))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(optimizer='rmsprop',loss="binary_crossentropy",metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,batch_size=300,epochs=10,verbose=1,validation_split=.2)
```

```python
model.evaluate(x_test,y_test)
```

```python
y_pred = model.predict_classes(x_test)
```

```python
history.history
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(["Train","Test"],loc="upper left")
plt.plot()
plt.show()

plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(["Train Loss","Test Loss"],loc="upper left")
plt.plot()
plt.show()
```
