```python
op = pd.read_csv('../input/one-piece-anime/ONE PIECE.csv')
op.head()
```

```python
op.tail()
```

```python
op = op.drop('Unnamed: 0', axis = 1)
```

```python
op.shape
```

```python
op.isnull().sum()
```

```python
op.describe().transpose()
```

```python
op.info()
```

```python
op.loc[~op['trend'].str.isdigit(), 'trend'].tolist()
```

```python
op['rank'] = op['rank'].str.replace(',', '')
op['trend'] = op['trend'].str.replace(',', '')
op['trend'] = op['trend'].str.replace('-', '0')
op['total_votes'] = op['total_votes'].str.replace(',', '')
```

```python
op['rank'] = op['rank'].astype('int')
op['trend'] = op['trend'].astype('int')
op['total_votes'] = op['total_votes'].astype('int')
```
