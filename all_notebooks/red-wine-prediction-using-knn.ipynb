```python
%matplotlib notebook
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

** **LETS IMPORT THE DATASET **

```python
wine=pd.read_csv('../input/red-wine-quality-cortez-et-al-2009/winequality-red.csv')
wine.head()
```

**before classification first we check is there any null data **

```python
wine.isnull().sum()
```

So no NaN values.  consider, if the quality is above or equal to 7 the wine is good, orelse bad. for this classification we consider column quality for label.

```python
wine['quality']=['good' if x>=7  else 'bad' for x in wine['quality']]
wine['quality'].head()
```

lets split the data into train set and test set

```python
X=wine.drop('quality', axis =1)
y=wine['quality']
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=0)
```

**CLASSIFICATION**

```python
knn=KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train,y_train)
```

```python
knn.score(X_test,y_test)
```

.8925 is accuracy of this classifier

```python
w=knn.predict([[12.8,0.3,0.74,2.6,0.095,9,28,0.9994,3.2,0.77,10.8
]])
```
