The code is available on [my Github](https://github.com/aboyker/convnet-document-classification) if you are interested. This project implements a convolutional neural network inspired by the paper by [Yoon Kim](https://arxiv.org/abs/1408.5882). This is an end-to-end implementation that required minimal effort. Everything is already configure for this MBTI classification problem (prediction of the relevant personality type according to the 100 first terms of the posts written by each individual).

To train the model, just download the repository, cd to the repository directory and on the terminal type

`python train.py`

This implementation requires a recent version of tensorflow.

With the default parameters, the results are displayed in this confusion matrix (computed on validation data).

This project can be used to many others text classification tasks, and has been tested on some benchmark datasets with great success (see  [repo](https://github.com/aboyker/convnet-document-classification) for more details  ). Please star if you like it!
