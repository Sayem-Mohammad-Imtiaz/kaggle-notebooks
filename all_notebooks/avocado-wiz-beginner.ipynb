**Conent**
 The table below represents weekly 2018 retail scan data for National retail volume (units) and price. Retail scan data comes directly from retailers’ cash registers based on actual retail sales of Hass avocados. Starting in 2013, the table below reflects an expanded, multi-outlet retail data set. Multi-outlet reporting includes an aggregation of the following channels: grocery, mass, club, drug, dollar and military. The Average Price (of avocados) in the table reflects a per unit (per avocado) cost, even when multiple units (avocados) are sold in bags. The Product Lookup codes (PLU’s) in the table are only for Hass avocados. Other varieties of avocados (e.g. greenskins) are not included in this table.

Some relevant columns in the dataset:

Date - The date of the observation
AveragePrice - the average price of a single avocado
type - conventional or organic
year - the year
Region - the city or region of the observation
Total Volume - Total number of avocados sold
4046 - Total number of avocados with PLU 4046 sold
4225 - Total number of avocados with PLU 4225 sold
4770 - Total number of avocados with PLU 4770 sold

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings("ignore", message="numpy.dtype size changed")

```

Loading the data 

```python
df=pd.read_csv("../input/avocado.csv")
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.drop("Unnamed: 0", axis=1, inplace= True)
df.head()
```

```python
# df.set_index("Date",inplace=True)
plt.figure(figsize=(15,25))
sns.scatterplot(x="Date", y="AveragePrice",data=df,hue="type",alpha=0.5)
```

#Seperating the date, month and year to a separate column 

```python

df["Year"]=df["Date"].apply(lambda time:time.year)
df["Month"]=df["Date"].apply(lambda time:time.month)
df.head()
```

```python
plt.figure(figsize=(18,15))
sns.lineplot(x="Year", y="AveragePrice", data=df, hue="type")
```

```python
plt.figure(figsize=(18,15))
sns.lineplot(x="Month", y="AveragePrice", data=df, hue="type")
```

```python
#Differnt Aavocados other than HASS 4046,4770,4225
df["Other_Av"]=df['Total Volume']-df['4046']+df['4225']+df["4770"]
df.head()
```

```python
#conventional Avocados --- Region wise Price change across years 
av_conv=df[df["type"]=="conventional"]
av_conv.head()
plt.figure(figsize=(18,15))
sns.lineplot(x="Year",y="AveragePrice", hue="region", data=av_conv)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.factorplot(x="AveragePrice",y="region",hue="year",data=df,size=20,aspect=0.8,join=False)
```

```python
#Region wise Average Pricing
df.head()
g_1=df.groupby(["region"],as_index=False).mean()
g_1=g_1.sort_values("AveragePrice",ascending=False)
plt.figure(figsize=(15,10))
plt.xticks(rotation=90)
sns.barplot(x="region",y="AveragePrice", data=g_1)

```

```python
g_2=df.groupby(["region","type"],as_index=False).mean()
sns.factorplot(x="AveragePrice",y="region",hue="type",data=g_2,height=20,aspect=0.8,join=True)

```

```python
corr_1=df.drop(["Year","year","Month"],axis=1).corr()
plt.figure(figsize=(15,15))
sns.heatmap(corr_1,annot=True)
```
