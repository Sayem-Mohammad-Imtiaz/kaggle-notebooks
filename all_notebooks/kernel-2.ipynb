```python
def f(*args):
    for i in args:
        print(i)
f(3)
print("")
f(1,2,3,4,5,6,7)
def f(**kwargs):
    for key, value in kwargs.items():
        print(key,"",value)
f(country='Sweden',capital='Stockholm',population='5344534')
```
