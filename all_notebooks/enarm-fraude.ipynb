```python
#Lista de estados de México

Estados_Mex = ['AGUASCALIENTES','BAJA CALIFORNIA','BAJA CALIFORNIA SUR','CAMPECHE','CHIHUAHUA',
              'CHIAPAS','COHALITA DE ZARAGOZA','COLIMA','DURANGO','GUANAJATO','GUERRERO',
              'HIDALGO','JALISCO','ESTADO DE MÉXICO','MICHOACÁN DE OCAMPO','MORELOS','NAYARIT','NUEVO LEÓN',
              'OAXACA','COAHUILA DE ZARAGOZA','QUERÉTARO','QUINTANA ROO','SAN LUIS POTOSÍ','SINALOA','SONORA',
              'TABASCO','TAMAULIPAS','TLAXCALA','VERACRUZ','YUCATÁN','ZACATECAS','MÉXICO']

#Lista de capitales de estado (Palabras clave no repetidas)

Cap_est_Mex = ['AGUASCALIENTES','MEXICALI','PAZ','FRANCISCO','CHIHUAHUA',
               'TUXTLA GUTIÉRREZ','SALTILLO','COLIMA','VICTORIA DE DURANGO','GUANAJUATO',
              'CHILIPANCINGO BRAVO','PACHUCA SOTO','GUADALAJARA',
               'TOLUCA DE LERDO','MORELIA','CUERNAVACA','TEPIC','MONTERREY','JUÁREZ',
              'PUEBLA','SANTIAGO','CHETUMAL','SSLLPP','CULIACÁN ROSALES',
              'HERMOSILLO','VILLAHERMOSA','CIUDAD VICTORIA','XICOHTÉNCATL','XALAPA-ENRÍQUEZ',
              'MÉRIDA','ZACATECAS','CIUDAD DE MÉXICO']
# Últimas palabras clave

Key_words=['ANAHUAC','ANÁHUAC','COAHUILA','TAMAULIPAS-TAMPICO','ZARAGOZA','IZTACALA','XOCHICALCO',
           'QUETZALCÓATL','QUETZALCOATL','CD. VICTORIA','MATATIPAC','XOCHIMILCO','VALLE DEL BRAVO',
           'VALLE DE BRAVO','VILLA RICA','INSTITUTO POLITÉCNICO NACIONAL','ESCUELA SUPERIOR DE MEDICINA',
          'IPN','I.P.N.','FUERZA AÉREA','NAVAL','OTRO PAÍS','EXTRANJEROS','EXTRANJERAS','EXTRANJERO',
          'TOMINAGA NAKAMOTO','WEST HILL','SALLE', 'UNAM','MAYAB', 'AUTÓNOMA','SURESTE','SIERRA','NORESTE',
          'TEPEACA','FACULTAD DE MEDICINA','HOMEOPATÍA','CIENCIAS DE LA SALUD']
Est_corr =['NUEVO LEÓN','NUEVO LEÓN','COAHUILA DE ZARAGOZA','TAMAULIPAS','ESTADO DE MÉXICO',
           'ESTADO DE MÉXICO','MORELOS','GUANAJATO','GUANAJATO','TAMAULIPAS','NAYARIT','ESTADO DE MÉXICO',
           'ESTADO DE MÉXICO','ESTADO DE MÉXICO','VERACRUZ','ESTADO DE MÉXICO','ESTADO DE MÉXICO',
           'ESTADO DE MÉXICO','ESTADO DE MÉXICO','OTRA','OTRA','EXTRANJERO','EXTRANJERO','EXTRANJERO',
           'EXTRANJERO','ESTADO DE MÉXICO','ESTADO DE MÉXICO','ESTADO DE MÉXICO','ESTADO DE MÉXICO',
           'YUCATÁN','ESTADO DE MÉXICO','OAXACA','ESTADO DE MÉXICO','TAMAULIPAS','ESTADO DE MÉXICO',
           'ESTADO DE MÉXICO','ESTADO DE MÉXICO','JALISCO']
```

```python
est=[]
for esc in df['Escuela de Medicina'].unique():
    est.append(df.estado[df['Escuela de Medicina'].isin([esc])].unique()[0])
    
Est_Esc=pd.DataFrame(list(zip(est,df['Escuela de Medicina'].unique())))

for i in range(0,df.shape[0]):
    df.at[i,'estado'] = Est_Esc.at[Est_Esc.index[Est_Esc[1].isin([df['Escuela de Medicina'][i]])==True][0],0]

print('Busqueda por Universidades que tienen el nombre del estado en algunos años')
print(df.info())
print('Faltan {} datos'.format(df.estado.isnull().sum()))
print()

null_items = df.index[df.estado.isnull()].values.tolist()
for item in null_items:
    if any([word in df['Escuela de Medicina'][item] for word in Estados_Mex]):
        df.at[item,'estado'] = [word for word in Estados_Mex if word in df['Escuela de Medicina'][item]][0]

print('Busqueda por Estado en el nombre de la universidad')
print(df.info())
print('Faltan {} datos'.format(df.estado.isnull().sum()))
print()

null_items = df.index[df.estado.isnull()].values.tolist()
for item in null_items:
    if any([word in df['Escuela de Medicina'][item] for word in Cap_est_Mex]):
        index = Cap_est_Mex.index([word for word in Cap_est_Mex if word in df['Escuela de Medicina'][item]][0])
        df.at[item,'estado'] = Estados_Mex[index]

print('Busqueda por capital de Estado en el nombre de la universidad')
print(df.info())
print('Faltan {} datos'.format(df.estado.isnull().sum()))
print()

null_items = df.index[df.estado.isnull()].values.tolist()
for item in null_items:
    if any([word in df['Escuela de Medicina'][item] for word in Key_words]):
        index = Key_words.index([word for word in Key_words if word in df['Escuela de Medicina'][item]][0])
        df.at[item,'estado'] = Est_corr[index]

print('Busqueda por Palabras clave en el nombre de la universidad')
print(df.info())
print('Faltan {} datos'.format(df.estado.isnull().sum()))
print()
```

Dataset completo.

```python
df.replace({'Escuela de Medicina' : {'UNIVERSIDAD':'UNIV.',
                                     'AUTÓNOMA':'AUT.',
                                     'INSTITUTO':'INST. ',
                                     'TECNOLÓGICO Y': 'TECNOLÓGICO',
                                     'TECNOLÓGICO DE': 'TECNOLÓGICO',
                                     'ESTUDIOS' : 'EST.',
                                     'SUPERIORES' : 'SUP.',
                                     'DE MONTERREY' : 'MTY.',
                                     'BAJA CALIFORNIA' :'B.C.',
                                     'ESCUELA' :'ESC.',
                                     'NAVAL' :'NAV.',
                                     
                                     }},
          inplace=True,regex=True)
```

```python
mean_promedio=[]
years = np.arange(2001,2020)
for year in years:
    mean_promedio.append(df.promedio[df.año.isin([year])].mean())
año_promedio =pd.DataFrame(list(zip(years,mean_promedio)),columns=['Año','Media Promedio'])
año_promedio.fillna(0,inplace=True)
año_promedio.set_index('Año',inplace=True)
año_promedio
```

```python
df['Varianza']=np.zeros(df.shape[0])
for year in years:
    df.loc[df.año.isin([year]),'Varianza']=df['promedio'][df.año.isin([year])]-año_promedio.at[year,'Media Promedio']
        
df.head()
```

```python
df.sort_values('Varianza',ascending=False).head(30)
```

Las 30 universidades con mayor varianza de la media, estas universidades y las subsiguientes serán susceptibles de fraude.

## Visualización y búsqueda de fraude

```python
fig = px.scatter(df, x='año', y='promedio', color='Escuela de Medicina',
             hover_name='Escuela de Medicina', hover_data=['estado','promedio','Sustentantes','Seleccionados','Varianza'])
fig.add_trace(go.Scatter(x=año_promedio.index,y=año_promedio['Media Promedio'],name='Media de la nota promedio'))

fig.update_layout(legend_orientation ='h',legend_y =-0.3,
                 title='Promedio de nota por año y universidad')
fig.show()
```

Se puede observar alguna universidad con un cambio drástico de nota promedio como la Universidad del Valle de Mexico -Campus Jalisco pero puede ser debido a que el número de sustentantes es reducido.

Hay que buscar picos que se salgan de la media de las universidades.

```python
fig = px.scatter(df.sort_values('Varianza',ascending=False)[0:100], x='año', y='Varianza', color='Escuela de Medicina',
             hover_name='Escuela de Medicina', hover_data=['estado','promedio','Sustentantes','Seleccionados'])

fig.update_layout(legend_orientation ='h',legend_y =-0.3,
                 title='100 Casos más susceptibles de fraude')
fig.show()
```

Se puede observar que la media es de:
* 40 puntos : 2001-2006 y 2008
* 30 puntos : 2007
* 50 puntos : 2011
* 60 puntos : 2012-2019
Esto da una idea de la dificultad del examen, para una universidad que enseña el mismo contenido cada año la varianza respecto a la media debería tender a ser constante. Una curva en forma de 'U' puede indicar el punto de inflexión a partir del cuál se comenzó a cometer fraude de forma continua, una curva en forma de 'A' indicaría que se dejó de hacer fraude, un pico sería fraude puntual, y una línea constante ausencia de fraude.
