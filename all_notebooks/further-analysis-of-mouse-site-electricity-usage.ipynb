```python
#imports
import pandas as pd
import numpy as np
```

```python
#load dataset 
electricity = pd.read_csv("../input/bdgp2-further-cleaned-datasets/electricity_cleaned_new.csv")
```

```python
electricity.head()
```

```python
#show types of the values 
electricity.dtypes
```

```python
#drop unnamed column 
electricity = electricity.drop("Unnamed: 0", axis = 1)
```

```python
#change to DateTime format
electricity["timestamp"] = pd.to_datetime(electricity["timestamp"], format = "%Y-%m-%d %H:%M:%S")
```

```python
#set time as index 
electricity = electricity.set_index("timestamp")
```

```python
#average by week
electricity = electricity.resample("W").mean()
```

```python
mouse_electricity = pd.DataFrame()
M = [col for col in electricity.columns if 'Mouse' in col]
mouse_electricity[M] = electricity[M]
```

## Analysis 
* Comparing the newest building to the oldest building (Mouse_health_Justin & Mouse_health_Estella): 
    * Both buildings use district heating 
    * Both buildings are used for healthcare
    * The older building (Mouse_health_Estella) has less yearly electricity usage per m^2 when compared to the newer building (Mouse_health_Justin)
    * But we should not be making this direct comparison because: 
        * As hospitals modernize and expand, there will be new technologies and machines that will require more electricity usage 
            * Electricity usage (excluding heating) accounts for a total of 21.46% of this sites total electricity usage in 2015
            * We do not know the improvements the hospital made for each building. E.g. Possible that one building received more development than the other. 
    * We do not know the building material 
         * By looking at images, we see that Mouse_health_Justin has many more windows than Mouse_health_Estella. But new modern windows now contain more panes of glass providing better insulation. 
             * Doors and windows: the number, size and quality of doors and windows can influence heat lose up to 25% [[1]](https://www.hydroquebec.com/residential/customer-space/electricity-use/electricity-consumption-factors.html)
         * Additionally, the insulation and buliding material used in each building is unknown. 
             * Thermal envelope: poorly insulated walls, foundations, roof spaces can cause up to lose of 40% of a building's heat
         * These variables affect heating, which accounts for 78.5% of this site's electricity usage.
