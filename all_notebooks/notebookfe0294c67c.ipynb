```python
import pandas as pd
import numpy as np

data = pd.read_csv('../input/piratesvsglobalwarmingcorrelationcausation/pirates-vs-global-warming.csv')

data
```

```python
data.corr()
```

```python
data.plot.scatter(x='global-average-temperature', y='number-of-pirates', marker='x');
```

## Linear model
Fitting a line to predict the number of pirates

```python
params = np.polyfit(data['global-average-temperature'], data['number-of-pirates'], 1)
params
```

```python
x = data['global-average-temperature']
data['predicted-pirates'] = x * params[0] + params[1] 
ax = data.plot.scatter(x='global-average-temperature', y='number-of-pirates')
data.plot(x='global-average-temperature', y='predicted-pirates', ax=ax, color='r');
rmse = ((data['predicted-pirates'] - data['number-of-pirates']) ** 2).mean() ** 0.5
print(f"rmse: {rmse}")
```

## Quadratic model
This might be a better fit, as the line seems a bit too simplistic (underfitted)

```python
params = np.polyfit(data['global-average-temperature'], data['number-of-pirates'], 2)
params
```

```python
data['predicted-pirates'] = data['global-average-temperature']**2 * params[0] + data['global-average-temperature'] * params[1] + params[2]
ax = data.plot.scatter(x='global-average-temperature', y='number-of-pirates')
data.plot(x='global-average-temperature', y='predicted-pirates', ax=ax, color='r');
rmse = ((data['predicted-pirates'] - data['number-of-pirates']) ** 2).mean() ** 0.5
print(f"rmse: {rmse}")
```

RMSE for quadratic is much lower, and also visually it seems like a better model.

The model looks like a set of lines at this point, rather than a continuous curve. 
To fix this, you need to start using the Matplotlib library directly (Pandas is using this under the hood), or you need to create a new dataset

```python
import matplotlib.pyplot as plt
x1 = np.linspace(13, 17, 100)
y1 = x1**2 * params[0] + x1*params[1] + params[2]
plt.plot(x1, y1)
plt.plot(data['global-average-temperature'], data['number-of-pirates'], 'x');
```
