Standard deviation: $\sigma_{x}=\sqrt{\frac{1}{n-1} \sum_{i=1}^{n} (x_i-m_{x})^2}$


Covariance: $cov_{x,y}=\frac{1}{n-1} \sum_{i=1}^{n} (x_i-m_x)(y_i-m_y)$


Correlation: $cor_{x,y}=\frac{cov_{x,y}}{\sigma_x\sigma_y}$.

Machine learning: linear regression

```python
from sklearn.model_selection import train_test_split
```

Splitting data into training data and testing data

```python
X=df['YearsExperience']
Y=df['Salary']
```

```python
result=train_test_split(X,Y,test_size=0.2) #20% of data will be used for testing 
```

```python
X_train=result[0] #predictors (features) for training 
X_test=result[1]  #predictors  (features) for testing
Y_train=result[2] #variable (labels) to be predicted used for training
Y_test=result[3]  #variables (labels) to be predicted used for testing
```

X_train, X_test, Y_test, Y_train are series

```python
X_train.shape
```

```python
type(X_train)
```

```python
Y_train.shape
```

Import the type LinearRegression()

```python
from sklearn.linear_model import LinearRegression
```

Create a variable storing the model, first the untrained one, then train it. 

```python
lr=LinearRegression() #untrained model
```

Train your model using data

```python
lr.fit(X_train.to_numpy().reshape(-1,1),Y_train)
```

Reformat the variable X_train: convert it to a numpy.array and reshap it

```python
xt=X_train.to_numpy()   #conversion to a numpy array and then use reshape(-1,1)
```

```python
lr.fit(xt.reshape(-1,1),Y_train) #train our linear regression model
```

Test our model

```python
xtest=X_test.to_numpy().reshape(-1,1)
```

```python
Ypred=lr.predict(xtest)  #use test data to predict the variable to be predicted
```

```python
Ypred-Y_test
```

Evaluate the error: use norms, relative  sum of square error

```python
np.linalg.norm(Ypred-Y_test)/np.linalg.norm(Y_test) #relative error
```

```python
np.mean(Ypred-Y_test) #average error
```

```python
np.sqrt(np.linalg.norm(Ypred-Y_test)**2/Ypred.shape[0]) # average square error
```

Relative error is quite small

```python
plt.plot(range(Ypred.shape[0]),Y_test,'r', range(Ypred.shape[0]),Ypred,'b')
#plt.legend('True value','Predicted value')
```
