<h1 style = "color:red; font-weight:bolder; text-decoration: underline">Non-Categorical Data</h1>

<h1 style = "color:brown">1) Using MinMaxScaler from sklearn.preprocessing library of python</h1>

To achieve the above goal, I will use sklearn.preprocessing library of python.

```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
df_values = df.values
df_valued = Scaler.fit_transform(df_values)
normalized_df = pd.DataFrame(df_valued)
normalized_df
```

Here, I have coverted all the values of all attributes in the range of 0 to 1. Let's check the minimum and maximum values.

```python
normalized_df.describe()
```

<h1 style = "color:brown">2) Using StandardScaler sklearn.preprocessing library of python</h1>

To convert our data into standard form i.e. standard deviation is 1 and mean value is approximately equal to 0.

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
df_values = df.values
df_std = std_scaler.fit_transform(df_values)
std_df = pd.DataFrame(df_std)
std_df
```

```python
std_df.describe()
```

I have converted data into standarized form.<br><br> We have work on the non-categorical data till now. Now, I will work on the categorical data.

<h1 style = "color:red; font-weight:bolder; text-decoration: underline">Categorical Data</h1>

To work on the categorical data, I will use another dataset, which having categorical data. I will read that csv file and print the content.

```python
df1 = pd.read_csv('../input/categorical-data/stdcat-201010-101522.csv')
df1
```

```python
df1.info()
```

Here, we can see, the type of all the attributes is object except RNo.

This dataset have categorical data. Let's discuss the methods to convert it into standard data.

<h1 style = "color:brown">1) Using map method</h1>

map() method takes the new dictionary of updated values, and we can change that categorical data into numeric data.

```python
new = {'F':0, 'M':1}
df1.Gender = df1.Gender.map(new)
df1
```

```python
df1.info()
```

Now, we can see that the type of Gender is changed to int64.

Here we go...
  I have converted the categorical data into numeric data. We can use replace() method also.

<h1 style = "color:brown">2) Using replace method</h1>

We can use this method as like map() method.

<h1 style = "color:brown">3) Using LabelEncoder from sklearn.preprocessing</h1>

Here, I will convert State attribute into the numeric data.

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
df1['State'] = lb.fit_transform(df1['State'])
df1
```

Here, I have converted State attribute into numeric. We can check as follows:

```python
df1.info()
```

<h1 style = "color:brown">4) Using get_dummies method</h1>

```python
df1 = pd.get_dummies(df1, columns = ['Category'], prefix = ['Cat'])
df1
```

It will create one column for each category. If that particular category is present, then it come with 1. Otherwise, 0.

```python
df1.info()
```
