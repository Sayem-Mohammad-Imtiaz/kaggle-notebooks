```python
df = pd.get_dummies(data = df,columns=["workclass","education","marital-status","occupation","relationship","race","sex","native-country"],drop_first=True)
```

```python
from sklearn.impute import KNNImputer
model = KNNImputer(n_neighbors=4)
df = pd.DataFrame(model.fit_transform(df),columns = df.columns)
df
```

```python
df.info()
```

```python
from sklearn import preprocessing

col=[]
for i in df.columns


x = df[col].vales#returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled,columns=df.columns)
```

# Data Prep Done

Explorations and handeling missing values is done for the dataset.Data ready for models now. 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(df.drop("income >50K",axis=1),df["income >50K"],test_size=0.25)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
accuracy_score(model.predict(X_test),y_test)
```

```python
model = RandomForestClassifier()
model.fit(X_train,y_train)
accuracy_score(model.predict(X_test),y_test)
```

```python
model = XGBClassifier()
model.fit(X_train,y_train)
accuracy_score(model.predict(X_test),y_test)
```

```python
model = KNeighborsClassifier(n_neighbors=4)
model.fit(X_train,y_train)
accuracy_score(model.predict(X_test),y_test)
```

```python
model = DecisionTreeClassifier(max_depth=3)
model.fit(X_train,y_train)
accuracy_score(model.predict(X_test),y_test)
```
