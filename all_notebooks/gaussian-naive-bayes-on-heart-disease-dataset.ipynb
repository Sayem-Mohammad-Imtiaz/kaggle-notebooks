```python
# GAUSSIAN NAIVE BAYES ALGORITHM ON 'heart_disease' DATASET TO COMPARE PREDICTED VALUE AND ORIGINAL VALUE AND CALCULATE ACCURACY
# WE WILL BE USING COLUMNS FROM 'age' to 'thal' FOR PREDICTION
# ACCURACY IS MORE IN THIS CELL AS IT INVOLVES ALL THE RELATABLE COLUMNS

import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn import model_selection

df=pd.read_csv('../input/heart-disease-patients-details/heart_disease.csv')

Y=df['result']
X=df.loc[:,'age':'thal']

X_train,X_test,Y_train,Y_test=model_selection.train_test_split(X,Y,random_state=7,test_size=0.4)

gb=GaussianNB()
gb.fit(X_train,Y_train)

print(gb.score(X_train,Y_train))

prediction=gb.predict(X_test)
print(prediction)
print(Y_test)
```
