#Features are `x_cols`

```python
# load data
train_data = pd.read_csv('../input/pokemon-datasets-for-ml/train_pokemon.csv')
test_data  = pd.read_csv('../input/pokemon-datasets-for-ml/train_pokemon.csv')
x_cols = ['Number', 'Name', 'Total', 'HP', 'Sp_Atk', 'Sp_Def', 'Speed', 'Generation','Color', 'hasGender', 'Egg_Group_1', 'hasMegaEvolution','Height_m', 'Weight_kg', 'Body_Style', 'shuffle']
```

#Explain Mode: for data exploration.The Automatic EDA report is created. Extensive explanations for models are computed.

```python
automl = AutoML(mode="Explain")
automl.fit(train_data[x_cols], train_data["isLegendary"])
```

```python
# check AutoML models
# you can click on links!
automl.report()
```

#COMPETE MODE - Golden Features and Features Selection are performed. Many algorithms are tuned, stacked and ensembled.

```python
automl = AutoML(mode="Compete", total_time_limit=4*3600)
automl.fit(train_data[x_cols], train_data["isLegendary"])
```

```python
# check AutoML models
# you can click on links!
automl.report()
```
