# K-Means Classifier
An unsupervised learning algorithm which either combines or seperates two data points on the basics of distance.


```python
## calling the kmeans from sklearn
from sklearn.cluster import KMeans
```

```python
## Giving no of cluster as  we already know 
## The outcome have only two catogeries
kmeans = KMeans(n_clusters=2)
```

As you can see our Data set contain Private column in advance, we need to drop it before/ while fitting the model

```python
## Fitting the model on whole data
## As it is supervised learning with no target splitting is unnecessary
kmeans.fit(df.drop('Private',axis=1))
```

```python
## Cluster centers
kmeans.cluster_centers_
```

**Convert Count Function:**

The Private column is a catogerical column we need to convert it into numerical column.There are several ways to do the following.

we are going to define a function to return 1 & 0

```python
## Convert Count Function:
def converter(private):
    if private=='Yes':
        return 1
    else:
        return 0
```

```python
## Applying the convert count function to our Private column &
## Creating Cluster column
df['Cluster'] = df['Private'].apply(converter)
```

```python
## Checking the dataframe
df.head()
```

# Accuracy 

We cannot check the accuracy of a unsupervised machine learning, but in our case we do have outcome to check how our model performed.

Let's Check how our Model performed.

```python
## Confusion Matrix & Classification report
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(df['Cluster'],kmeans.labels_))
print(classification_report(df['Cluster'],kmeans.labels_))
```
