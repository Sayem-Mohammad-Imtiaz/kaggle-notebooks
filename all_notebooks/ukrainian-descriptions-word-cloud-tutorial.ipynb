# Word Cloud for [Ukrainian Descriptions Of Words](https://www.kaggle.com/ihelon/ukrainian-descriptions-of-words) dataset

This is the guide about using wordcloud library for building images from words.   
We will use Ukrainian Descriptions Of Words dataset, which contains a lot of ukrainian words. But you can use your own text corpus in other languages.

![](https://i.imgur.com/UqXawPb.png)

```python
import urllib
import random

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from wordcloud import WordCloud, ImageColorGenerator
```

## Load data

Let's load the dataset with text and convert it into one text corpus

```python
path_data = "../input/index.csv"
df = pd.read_csv(path_data)

text = ' '.join(df["description"].values)
print("Text example: ")
text[:300]
```

## Minimal example of the word cloud

```python
wordcloud = WordCloud(
    background_color="black",
    width=500,
    height=500,
    random_state=42,
)

wordcloud.generate(text)


plt.figure(figsize=(5, 5))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

## Create word cloud from mask

 Load simple mask as numpy array (you can use your own image)

```python
def load_mask(mask_url):
    with urllib.request.urlopen(mask_url) as resp:
        mask = np.asarray(bytearray(resp.read()), dtype="uint8")
        mask = cv2.imdecode(mask, cv2.IMREAD_COLOR)
        mask = cv2.cvtColor(mask, cv2.COLOR_BGR2RGB)
        
    return mask


mask = load_mask("https://i.imgur.com/nY3pUfe.png")

plt.imshow(mask)
plt.axis("off");
```

Create and show default wordcloud (with the mask)

```python
wordcloud = WordCloud(
    background_color="white", 
    mask=mask,
    max_words=500,
    random_state=42,
)

wordcloud.generate(text)


plt.figure(figsize=(8, 8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

We can control percentage of horizontal and vertical word arrangements using parameter *prefer_horizontal*. Set it to 0.1 to see nearly all words in vertical  arrangements

```python
wordcloud = WordCloud(
    background_color="white", 
    mask=mask,
    max_words=500,
    random_state=42,
    prefer_horizontal=0,
)

wordcloud.generate(text)


plt.figure(figsize=(8, 8))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

## Colormaps for word cloud

We can use different color maps for the clouds

```python
colormaps = [
    "winter", "spring", "summer", 
    "autumn", "gray", "rainbow", 
    "Blues", "jet", "Oranges",
]

plt.figure(figsize=(16, 16))

for ind, colormap in enumerate(colormaps, 1):
    wordcloud = WordCloud(
        background_color="white", 
        mask=mask,
        max_words=200,
        colormap=colormap,
        random_state=42,
    )
    
    wordcloud.generate(text)

    plt.subplot(3, 3, ind)
    plt.imshow(wordcloud)
    plt.title(f'colormap: {colormap}', fontsize=16)
    plt.axis("off")
```

## Colored masks

Let's load another colored mask as numpy array (you can use your own colored image)

```python
colored_mask = load_mask("https://i.imgur.com/HR1wh2N.png")

plt.imshow(colored_mask)
plt.axis("off");
```

We can color the wordcloud using an image-based coloring strategy implemented in ImageColorGenerator. It uses the average color of the region occupied by the word in a source image.

```python
wordcloud = WordCloud(
    background_color="white", 
    mask=colored_mask,
    max_words=500,
    random_state=42,
)
    
wordcloud.generate(text)

image_colors = ImageColorGenerator(colored_mask)

plt.figure(figsize=(8, 8))
plt.imshow(wordcloud.recolor(color_func=image_colors))
plt.axis("off")
plt.show()
```

## Save word cloud to file

To save wordcloud we can simply use wordcloud.to_file(path) method

```python
wordcloud.to_file("example.png")
```
