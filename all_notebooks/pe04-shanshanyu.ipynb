```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
```

```python
# Loading dataset
data = pd.read_csv("../input/did-it-rain-in-seattle-19482017/seattleWeather_1948-2017.csv")
data.head()
```

I will drop the date column because it's hard to vectorizing date data

```python
# Data processing
data = data.dropna(axis=0)
data = data.iloc[:, 1:5]
```

```python
data.head()
```

# Plotting All Data Points

```python
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

ax.set_xlabel("TMIN")
ax.set_ylabel("TMAX")
ax.set_zlabel("PRCP")
ax.scatter3D(xs=data["TMIN"],ys=data["TMAX"],zs=data["PRCP"], c=data["RAIN"])
plt.show()
```

2.	Choose 70% of your dataset as the training data for your K-Nearest Neighbor algorithm. 

```python
# 2. Spliting Dataset
train_set, test_set = train_test_split(data, train_size=0.7, test_size=0.3)
```

```python
train_set.head()
```

```python
test_set.head()
```

```python
train_data, train_lable = train_set.iloc[:, 0:3], train_set["RAIN"].astype('int')
test_data, test_lable = test_set.iloc[:, 0:3], test_set["RAIN"].astype('int')
```

```python
train_data
```

3.	Using the scikit-learn API, apply the K-Nearest Neighbor to your training set for 5 different values of K. 

```python
# KNN classifier
from sklearn.neighbors import KNeighborsClassifier
Knn_classifiers = []

# k = 1, 3, 5, 7, 9
for k in range(1, 10, 2):
    classifier = KNeighborsClassifier(n_neighbors=k).fit(train_data, train_lable)
    Knn_classifiers.append(classifier)
```

4.	Use the additional 30% of your data as a testing dataset to predict which class they belong to. How does the number of nearest neighbors affect this prediction? 

```python
# Evaluating 5 classifiers with different k values
k = 1
for classifier in Knn_classifiers:
    print("K =", k, ", score =", classifier.score(test_data, test_lable))
    k += 2
```

Answer to 4: The lager K value, the lower accuracy of the classifier.

5.	Find the best results you can achieve using the K-Nearest Neighbor algorithm and interpret the obtained results based on the problem you are trying to solve. 

```python
# k = 1
predictions = Knn_classifiers[0].predict(test_data)
predictions
```

```python
test_data["RAIN"] = predictions.astype('bool')
```

```python
test_data.head()
```

# Plotting Testing-Predictions


```python
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")

ax.set_xlabel("TMIN")
ax.set_ylabel("TMAX")
ax.set_zlabel("PRCP")
ax.scatter3D(test_data["TMIN"], test_data["TMAX"],test_data["PRCP"], c=test_data["RAIN"])
plt.show()
```
