```python
import pandas as pd
import numpy as np
import sklearn
from sklearn import linear_model
from sklearn.utils import shuffle

print('Imports good')
```

```python
#data path
dataPath = "../input/acamedicperfomance/student_math.csv"
data = pd.read_csv(dataPath, sep=";")
print("data yes")
```

```python
#dataframe
print(data.head())
```

```python
data = data[["G1", "G2", "G3", "studytime", "failures", "absences"]]
```

```python
predict = "G3"

x = np.array(data.drop([predict], 1))
y = np.array(data[predict])
```

```python
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size = 0.1)
```

```python
linearRegress = linear_model.LinearRegression()
```

```python
linearRegress.fit(x_train, y_train)

accuracy = linearRegress.score(x_test, y_test)
```

```python
predictions = linearRegress.predict(x_test)

for x in range(len(predictions)):
    print(predictions[x], x_test[x], y_test[x])
```

```python
print(accuracy*100)
```
