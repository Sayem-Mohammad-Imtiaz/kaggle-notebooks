Import Libraries

```python
import math
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.rcParams["figure.figsize"] = (20,10)

import missingno as msno

import warnings
warnings.filterwarnings("ignore")
```

```python
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#Read csv file
df = pd.read_csv('/kaggle/input/pakistans-top-250-youtube-channels/Pakistan Top 250 YouTube Channels.csv')
```

```python
#Check Shape
df.shape
```

```python
#Get 5 sample records across the entire dataset
df.sample(5)
```

```python
#Remove the first column.
df1 = df.drop(['Top #'], axis='columns')
df1.shape
```

```python
#The first column should be renamed
df2 = df1.rename(columns = {df1.columns[0]:'Channel'})
```

```python
#Separated column by commas
df3 = df2["Channel"].str.split(",",expand = True) 
df3.sample(5)
```

```python
#Check information of df3
df3.info()
```

```python
df4 = df3.rename(columns = {df3.columns[0]:'Channel', df3.columns[1]:'Category', df3.columns[2]:'Subscriber'})
df4.sample(5)
```

```python
#Check for missing values, but there are none in this dataset.
msno.matrix(df4)
```

```python
#Remove the last character from the previous column and position it in the current one.
df4["millinames"] = df4["Subscriber"].str[-1:]
df4["Subscriber"] = df4["Subscriber"].str[:-1]
#Object into Numeric
df4["Subscriber"] = pd.to_numeric(df4["Subscriber"], errors = 'coerce')
df4.sample(5)
```

```python
for index, row in df4.iterrows():
    if row["millinames"] == "M":
        df4.at[index, "Subscriber"] = df4.loc[index,"Subscriber"]*1000000
    elif row["millinames"] == "K":
        df4.at[index, "Subscriber"] = df4.loc[index,"Subscriber"]*1000
```

```python
df5 = df4.drop(["millinames"], axis='columns')
```

### Visualization

##### Which category has the most subscribers?

```python
sr = df5.groupby('Category')['Subscriber'].agg('sum')
dict1 = sr.to_dict()
k = dict1.keys()
v = dict1.values()

labels = 'Entertainment', 'Music', 'News & Politics', 'Education','People & Blogs', 'Film & Animation', 'Howto & Style', 'Sports','Science & Technology', 'Gaming', 'Comedy', 'Autos & Vehicles','Travel & Events', 'Nonprofits & Activism'
sizes = v

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax.axis('equal')
fig.set_size_inches(16,10)

plt.show()
```

```python
sr = df5.groupby('Category')['Subscriber'].agg('sum')
dict1 = sr.to_dict()
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0,0,1,1])
k = dict1.keys()
v = dict1.values()
ax.bar(k,v)
plt.xticks(rotation='vertical')
plt.show()
```

```python
#Values are converted to thousands, millions, billions, and trillions.
def millify_by_records(tax_record):
    millnames = ['','K','M','B','T']
    n = float(tax_record)
    millidx = max(0,min(len(millnames)-1,
                        int(math.floor(0 if n == 0 else math.log10(abs(n))/3))))
    m = round(n / 10**(3 * millidx))
    return str(m)+str(millnames[millidx])
```

##### Below is a visual representation of the top 50 most subscribed networks.

```python
df6 = df5.sort_values(by=['Subscriber'], ascending=False)
fig = plt.figure(figsize=(12,6))
ax = fig.add_axes([0,0,1,1])
k = df6["Channel"][:50]
v = df6["Subscriber"][:50]
bars = ax.bar(k,v)
plt.xticks(rotation='vertical')

for bar in bars:
  height = bar.get_height()
  label_x_pos = bar.get_x() + bar.get_width() / 2
  ax.text(label_x_pos, height, s=millify_by_records(height), ha='center', va='bottom')

plt.show()
```
