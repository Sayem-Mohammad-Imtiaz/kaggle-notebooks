```python
class util():
    def __init__(self,bpath=''):
        self.bpath = '/kaggle/input/'
        self.tg = 'custom_1988_2020'

        print(dt.now())
    # 
    def re(self,xlist,reg='^(?!^__)'):
        return([x for x in xlist if re.search(reg,x)])
    # Measuring Execution Time
    def tc(self,tg,rt=''):
        start_time = time.perf_counter()
        exec(tg)
        execution_time = time.perf_counter() - start_time
        print(execution_time)
        
        if len(rt) > 0:
            return(eval(rt))
        else:
            return
    # override function
    def ride(self,o, i):
        for k in o:
                if (k not in i.keys() or
                   type(o[k]) != dict or
                   type(i[k]) != dict):
                    i[k] = o[k]
                else:
                    i[k] = self.ride(o[k], i[k])
        return(i)
    
    def csv_read(self):
        f = open(self.bpath + self.tg + '.csv','r')
        xlist = []
        start = 0
        num = 113607322
        raw = f.readline().rstrip()
        start = start + len(raw) + 2
        ym_a = raw.split(',')[0]
        count = 1
        for i in range(113607322):
            raw = f.readline().rstrip()
            ym = raw.split(',')[0]
            if ym != ym_a:
                xlist.append([ym_a,count,start])
                ym_a = ym
                count = 1
            count +=1
            start = start + len(raw) + 2
        xlist.append([ym_a,count,start])
        f.close()
        


        return(xlist)
            
            

            

ut = util()
```
