```python
#Visualize whole data
import numpy as np
from scipy.stats import chi2_contingency 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Histogram
numbers = pd.Series(diabetes.columns)
diabetes[numbers].hist(figsize = (14, 14))
plt.show();
```

```python
diabetes.plot(kind = 'box', subplots = True, layout = (4, 4), sharex = False, sharey = False, figsize = (12, 12));
```

```python
#densityPlot
diabetes.plot(kind = 'density', subplots= True, layout = (4,4), sharex =False, figsize = (12, 12));
```

```python
#Correlation Heatmap
corr = diabetes.corr()
f, ax = plt.subplots(figsize = (20, 9))
sns.heatmap(corr, vmax = 8, annot = True);
```

```python
import plotly.express as px
fig = px.scatter(diabetes.query("Outcome==1"), x = "Glucose", y = "BloodPressure", size = "Insulin", color = "Pregnancies", hover_name = "Pregnancies", log_x = True, size_max = 60)
fig.show()
```
