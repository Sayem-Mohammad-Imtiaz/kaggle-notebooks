**Kiva Microfunds** (commonly known by its domain name, Kiva.org) is a 501(c) non-profit organization that allows people to lend money via the Internet to low-income entrepreneurs and students in over 80 countries. Kiva's mission is “to connect people through lending to alleviate poverty.”

Datasets:
1. Kiva Loans - This is a subset of Kiva's data snapshots which can be found on their API webpage. It contains data about some of Kiva's loans. Click on the Column Metadata tab to see column descriptions.
2. kiva_mpi_region_locations.csv - This file contains Kiva’s estimates as to the geolocation of subnational MPI regions which can be found in the OPHI’s datasets.
3. loan_theme_ids.csv - This file contains records from the Kiva Data Snapshot and can be matched to the loan theme regions to get a loan’s location.
4. loan_themes_by_region.csv - This file contains regional data related to loan themes and partner ids. They are Kiva’s estimates as to the various geolocations in which a loan theme has been offered, as well as the resulting estimate of which MPI Region(s) the loan theme is in.

```python
#Read input datasets:
kiva_loans = pd.read_csv("../input/kiva_loans.csv")
kiva_mpi_reg_locs = pd.read_csv("../input/kiva_mpi_region_locations.csv")
loan_theme_ids = pd.read_csv("../input/loan_theme_ids.csv")
loan_themes_by_reg = pd.read_csv("../input/loan_themes_by_region.csv")
```

```python
# Lets explore each dataset.
kiva_loans.head()
kiva_loans = kiva_loans.fillna('')
```

```python
#Lets see how is the loan fund distributed across Countries.
Kiva_loans_group_by_country=kiva_loans.groupby('country')['loan_amount'].sum().reset_index()
Kiva_loans_group_by_country=Kiva_loans_group_by_country.sort_values('loan_amount').reset_index()
```

```python
Kiva_loans_group_by_country.head()
```

```python
plt.figure(figsize=(12,20))
ax = sns.barplot(y="country", x="loan_amount", data=Kiva_loans_group_by_country)
```

**from above graph , Philippines has major share of loan.**

```python
#Lets see how is the loan fund distributed across Sectors.
Kiva_loans_group_by_sector=kiva_loans.groupby('sector')['loan_amount'].sum().reset_index()
Kiva_loans_group_by_sector=Kiva_loans_group_by_sector.sort_values('loan_amount').reset_index()
```

```python
plt.figure(figsize=(12,12))
ax = sns.barplot(y="sector", x="loan_amount", data=Kiva_loans_group_by_sector)
```

No Surprise, Agriculture being the most funded sector and it looks like more country side (villages) having fund requirement.

```python
#Wordcloud for Use Text
from wordcloud import WordCloud, STOPWORDS
Alltext= (' '.join(kiva_loans['use']))
wc = WordCloud(width = 1000, height = 500,stopwords=STOPWORDS).generate(Alltext)

plt.figure(figsize=(15,5));
plt.imshow(wc);
plt.axis('off');
plt.title('Word Cloud for Use text');
```

Looking at the word cloud we can see, borrowers are taking loan for business, selling, buying , drinking water purpose.
