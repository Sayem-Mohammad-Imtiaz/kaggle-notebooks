**Sobre os dados**

Os dados foram obtidos por meio de sites web, como por exemplo https://www.transfermarkt.pt/ , que apresenta informações como tabela de classificação, valor de um time naquele ano, quantidade de jogadores, entre outras. O dataset disponibilizado traz informações do Campeonato Brasileiro Série A dos anos de 2009 a 2018, bem como a classificação de um time em determinado ano, a quantidade de vitórias, derrotas, empates, idade média, o valor total dos jogadores, dentre outros.

Layou Explicativo da base de dados.
* **Ano**	- Ano da Edição do Campeonato 
* **Pos.** - Posição do Time no Ano
* **Clubes** - Nome do Time
* **Vitorias** - Numero de Vitórias
* **Derrotas** - Numero de Derrotas
* **Empates**	- Numero de empates
* **GolsF/S** - Numero de gols feitos e gols sofridos
* **Saldo**	- Saldo de Gols (Gols Feitos - Gol Sofridos)
* **Qtd_Jogadores**	- Quantidade de Jogadores
* **Idade_Media**	- Idade Media dos Jogadores
* **Estrangeiros** - Numero de jogadores Estrangeiros
* **Valor_total** - Valor total do passe de todos os jogadores do clube
* **Media_Valor** - Valor médio do passe dos jogadores do clube

# Fazendo a leitura dos dados

```python
dados = pd.read_csv('/kaggle/input/campeonato-braileiro-20092018/tabelas/Tabela_Clubes.csv')
```

# Analise inicial dos dados


```python
dados
```

# Deletando as ultimas 4 colunas, pois estão vazias.

```python
dados = dados.drop('Unnamed: 13', 1)
dados = dados.drop('Unnamed: 14', 1)
dados = dados.drop('Unnamed: 15', 1)
dados = dados.drop('Unnamed: 16', 1)
```

# Analisando os dados apos a exclusão das colunas vazias

```python
dados.head()
```

# Considerando que existem 20 times e o dados são de 10 anos, isso dá o total de 200 registros
* As informações abaixo mostra que temos 200 registros e que todos os campos estão preenchidos.

```python
dados.info()
```

# Alterar o tipo de dados da coluna Valor_total e Media_valor, para float

```python
dados['Valor_total'] = dados['Valor_total'].astype("float")
dados['Media_Valor'] = dados['Media_Valor'].astype("float")
dados['Ano'] = dados['Ano'].astype("float")
dados['QTD'] = 1
dados.info()
```

# Verificar se existe algum registro duplicado

```python
dados[dados.duplicated(keep=False)]
```

# Renomear uma coluna, para ficar claro do que se trata o dado.

```python
dados = dados.rename(columns={'Pos.': 'Posicao'})
```

# Dados apos renomear a coluna

```python
dados.info()
```

# Como podem ver,  a informação de gols feitos e sofridos estão na mesma coluna, separado por ":", vamos dividir essa coluna. 

```python
dados['GolsF/S']
```

# Dividindo a informação de gols feitos e gols sofridos.

```python
divisao=dados['GolsF/S'].str.split(':')
```

# Verificar como ficou a coluna dividida
* Nesse primeiro momento foi criado outro dataframe com as colunas divididas

```python
divisao.head()
```

# Atribuir nomes nas colunas que foram divididas

```python
GolsFeitos = divisao.str.get(0)
GolsSofridos = divisao.str.get(1)
```

# Adicionar as colunas no data frame

```python
dados['GolsFeitos']=GolsFeitos
dados['GolsSofridos']=GolsSofridos
```

# Alterar o tipo de dados das colunas que foram adicionadas

```python
dados['GolsFeitos'] = dados['GolsFeitos'].astype("int")
dados['GolsSofridos'] = dados['GolsSofridos'].astype("int")
```

# Verificar como ficou o dataframe apos a inclusao das novas colunas

```python
dados.head()
```

```python
dados.info()
```

# Deletar a coluna que usamos para dividir a informação de gols

```python
dados = dados.drop('GolsF/S', 1)
```

# Os dados estão prontos para analises

```python
dados.head()
```

# O Ano de cada edição está errado na base, está sempre um ano anterior, abaixo vou somar um ano para cada edição

```python
dados['Ano'] = dados['Ano']+1
```

```python
dados
```

# Sabendo que para cada ano existe uma classificação entre os times, vamos ver qual time ganhou mais campeonatos
* Para isso, vamos filtrar apenas os times que ganharam o titulo

```python
clubes_vencedores=dados[dados['Posicao']==1]
clubes_vencedores1=dados[dados['Posicao']==1]

clubes_vencedores
```

# Agregar os dados
* Em seguida vamos ordernar pelo maior numero de titulos


```python
clubes_vencedores=clubes_vencedores.groupby('Clubes')['Posicao'].sum().reset_index()
clubes_vencedores=clubes_vencedores.sort_values(by='Posicao', ascending=False)

clubes_vencedores1=clubes_vencedores1.groupby('Clubes')['Valor_total'].sum().reset_index()
clubes_vencedores1=clubes_vencedores1.sort_values(by='Valor_total', ascending=False)
```

# Grafico de barras para analisar melhor o resultado
* No grafico abaixo mostramos a quantidade de titulos por clube

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Clubes', y='Posicao', data=clubes_vencedores)
```

# Verificar o valor dos jogadores dos times que foram nos anos que foram campeões.

```python
sns.barplot(x='Clubes', y='Valor_total', data=clubes_vencedores1)
```

# Criação de divesos dataframes para graficos.

```python
clubes_valores1=dados
clubes_valores=dados
saldo_gols=dados
gols_times=dados
vice=dados
vice=vice[vice['Posicao']==2]
participacao=dados

#clubes_valores=dados[dados['Posicao']==1]
clubes_valores1=clubes_valores1.groupby('Clubes')['Valor_total'].sum().reset_index()
clubes_valores1=clubes_valores1.sort_values(by='Valor_total', ascending=False)
clubes_valores=clubes_valores.groupby('Clubes')['Valor_total'].sum().reset_index()
clubes_valores=clubes_valores.sort_values(by='Valor_total', ascending=False)
saldo_gols=saldo_gols.groupby('Ano')['GolsFeitos'].sum().reset_index()
saldo_gols=saldo_gols.sort_values(by='GolsFeitos', ascending=False)
gols_times=gols_times.groupby('Clubes')['GolsFeitos'].sum().reset_index()
gols_times=gols_times.sort_values(by='GolsFeitos', ascending=False)
vice=vice.groupby('Clubes')['QTD'].sum().reset_index()
vice=vice.sort_values(by='QTD', ascending=False)
participacao=participacao.groupby('Clubes')['QTD'].sum().reset_index()
participacao=participacao.sort_values(by='QTD', ascending=False)
```

 # Investimentos dos times nos 10 anos de análise
