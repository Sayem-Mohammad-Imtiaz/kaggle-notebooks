Hello People! Here we will try to predict the gender using SVM. Here is the flow of the notebook

1. Data Reading
2. Convert the target variable label into numeric variable
3. Model the data using SVM using different kernels
4. Using train and test classifier and measure the accuracy.
5. Select the best features only.
6. Find the best parameters using the GridSearch

I was able to achieve the accuracy of around 96.5% on an unseen data.

```python
#Read the data
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

dataset = pd.read_csv("../input/voice.csv")
# Any results you write to the current directory are saved as output.
```

```python
#Display the data and it's shape
#Display all the columns
pd.set_option("display.max_columns", 21)
print(dataset.head())
print(dataset.shape)
```

```python
# Target variable is categorical. We convert that into numeric. male : 0, female : 1
dict_gender = {
    "male": 0,
    "female": 1
}
dataset["label"] = dataset["label"].replace(dict_gender)
```

```python
#Find out the numeric and non numeric columns

numerics = ['int32', 'int64', 'float32', 'float64','int16','float16']

dataset_numeric = dataset.select_dtypes(include = numerics)
```

```python
dataset_numeric.shape
```

# We have checked that all the columns are numeric

 Let's find the correlation between the target variable and the features

```python
corr = dataset.corr()
corr["label"]
```

#There are some highly correlated features like meanfun. Let's use all the features for now  in our model and see the results

```python
#SVM Model without any paramter tuning and feature engineering
from sklearn import svm
from sklearn.metrics import accuracy_score
model = svm.SVC()
X_dataset = dataset.iloc[:,0:20]
y_dataset = dataset["label"]

model.fit(X_dataset, y_dataset)
predictions = model.predict(X_dataset)
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_dataset,predictions)
accuracy
 
```

# We got the accuracy of 76%. Let's change the kernel and see the results

```python
#SVM Model without any paramter tuning and feature engineering
from sklearn import svm
from sklearn.metrics import accuracy_score
model = svm.SVC(kernel = 'linear')
X_dataset = dataset.iloc[:,0:20]
y_dataset = dataset["label"]

model.fit(X_dataset, y_dataset)
predictions = model.predict(X_dataset)
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_dataset,predictions)
accuracy
```

# We got tremendous increase in the accuracy.

# Let's apply the model over different train and test data by using cross validation train test split

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
X_dataset = dataset.iloc[:,0:20]
y_dataset = dataset["label"]
X_train, X_test, y_train, y_test = train_test_split(
   X_dataset, y_dataset, test_size=0.33, random_state=1)
model = svm.SVC(kernel = 'linear')
model.fit(X_train, y_train)
predictions = model.predict(X_test)
accuracy = accuracy_score(y_test,predictions)
accuracy
```

#Our model performed brilliantly on unseen data. Now Let's see if selecting some specific features improve the accuracy or not.

```python
#Let's' use only 15 best features and observe the results
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
dataset_new = SelectKBest(chi2, k=15).fit_transform(X_dataset, y_dataset)
length = dataset_new.shape[1]
X_dataset_new = dataset.iloc[:,0:length-1]
y_dataset_new = dataset["label"]
X_train, X_test, y_train, y_test = train_test_split(
   X_dataset_new, y_dataset_new, test_size=0.33, random_state=1)
model = svm.SVC(kernel = 'linear')
model.fit(X_train, y_train)
predictions = model.predict(X_test)
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,predictions)
accuracy
```

#Let's perform grid search to find the best parameters:


```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
model = svm.SVC(kernel = 'linear')
X_dataset = dataset.iloc[:,0:20]
y_dataset = dataset["label"]
X_train, X_test, y_train, y_test = train_test_split(
   X_dataset, y_dataset, test_size=0.33, random_state=1)
parameters = {
    "C":[1.0, 2.0, 3.0],
    "decision_function_shape":["ovr", "ovo"]
}
grid = GridSearchCV(model,parameters)
grid.fit(X_train, y_train)
predictions = grid.predict(X_test)
print(grid.best_params_)
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,predictions)
accuracy
```

```python
print(grid.best_params_)
```

# Accuracy has increased. Now we have got an accuracy of 96.5%.

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
model = svm.LinearSVC()
X_dataset = dataset.iloc[:,0:20]
y_dataset = dataset["label"]
X_train, X_test, y_train, y_test = train_test_split(
   X_dataset, y_dataset, test_size=0.33, random_state=1)
parameters = {
    "C":[1.0,2.0,3.0],
    "multi_class":["ovr", "crammer_singer"]
}
grid = GridSearchCV(model,parameters)
grid.fit(X_train, y_train)
predictions = grid.predict(X_test)
print(grid.best_params_)
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,predictions)
accuracy
```
