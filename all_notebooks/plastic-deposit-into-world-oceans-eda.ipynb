```python
plastic=pd.read_csv('/kaggle/input/plastic-input-into-worlds-oceans/plastics-top-rivers.csv')
```

#### **Variables:**
#### * **Entity:** Specifies the river that emits plastic to the ocean
#### * **Country:** Country in which the river flows through
#### * **sharePlastics:** Share of global plastics emitted to ocean (%)

```python
plastic.head()
```

```python
plastic.info()
```

```python
# checking for null values
plastic.isnull().sum()
```

```python
# checking for duplicates
duplicate = plastic[plastic.duplicated()]
duplicate
```

```python
# total number of countries
plastic['Country'].nunique()
```

```python
# countrywise number of rivers flowing
df=plastic['Country'].value_counts().reset_index().rename(columns={'Country':'river_count','index':'Country'})
df.head()
```

```python
# top 5 rivers which has highest emission
plastic.sort_values('sharePlastics',ascending=False).head(5)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('dark_background')
```

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.pointplot(x='Country',y='river_count',data=df,color='aqua')
```

#### > Philippines has highest number of rivers flowing which emits plastics in the ocean.
#### > India stands at second position

```python
# checking outliers
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.boxplot(x='Country',y='sharePlastics',data=plastic,color='yellow')
```

```python
df1=plastic.groupby('Country').sum()
df1.head()
```

```python
# top 5 countries with highest share of global plastics emitted to ocean in %
df1.sort_values('sharePlastics',ascending=False).head(5)
```

```python
df1.sharePlastics.sum()
```

#### > 26.9945% of global plastics are emitted to ocean 

```python
colors = ['lightskyblue', 'red', 'blue', 'green', 'gold','aqua','purple','orange']
df1.sharePlastics.plot.pie(autopct='%1.2f%%',figsize=(8,8), label='emission', colors=colors)
plt.show()
```

#### > Philippines has highest percentage of plastic emission thorugh rivers in the ocean.
#### > India stands at 2nd place and Malaysia at 3rd
