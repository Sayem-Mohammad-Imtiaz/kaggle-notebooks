# SVM Classifier

```python
from sklearn import svm
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import average_precision_score

svm_classifier = svm.SVC()
svm_classifier.fit(x_train2, y_train2)

svm_classifier_pred = svm_classifier.predict(x_test)

print("Training Accuracy :", svm_classifier.score(x_train2, y_train2))
```

# Decision Tree Classifier

```python
from sklearn.tree  import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import average_precision_score

dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(x_train2, y_train2)

dt_classifier_pred = dt_classifier.predict(x_test)

print("Training Accuracy :", dt_classifier.score(x_train2, y_train2))
```
