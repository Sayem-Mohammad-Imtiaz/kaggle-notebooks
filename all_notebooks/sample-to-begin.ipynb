```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('../input/social-network-ads/Social_Network_Ads.csv')
df.head()
```

```python
df.Purchased.value_counts()
```

### NOW MAKE THE DEPENDENT MATRIX AND INDEPENDENT MATRIX

```python
X=df[['Age','EstimatedSalary']]
y=df['Purchased']
print(X.head())
y.head()
```

### NOW SPLIT THE DATASET INTO TRAINING SET AND TEST SET

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=23)
```

### TRAIN THE MODEL

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(X_train,y_train)
```

```python
result=model.predict(X_test)
print(result)
```

```python
score=100*model.score(X_test,y_test)
print(score)
```

### REPORT

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,result))
```
