```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv("../input/flight-take-off-data-jfk-airport/M1_final.csv")
print(df)
```

```python
df.info()
df.columns
#df=df.drop("TAIL_NUM",axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df.OP_UNIQUE_CARRIER=le.fit_transform(df.OP_UNIQUE_CARRIER)
df["DEST"]=le.fit_transform(df["DEST"])
df["Wind"] =le.fit_transform(df["Wind"].astype(str))
df["Condition"] =le.fit_transform(df["Condition"])
```

```python
X=df["TAXI_OUT"].values
Y=df.drop("TAXI_OUT",axis=1).values
from sklearn.model_selection import train_test_split
X_train,x_test,Y_train,y_test=train_test_split(X,Y,test_size=.1)
print(np.shape(X_train),np.shape(x_test),np.shape(Y_train),np.shape(y_test))
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

reg=LinearRegression()
reg=reg.fit(Y_train,X_train)

rmse1=(mean_squared_error(x_test,reg.predict(y_test))**0.5)
print(rmse1)
```

# Lasso Regression


```python
from sklearn import linear_model
lassoreg=linear_model.Lasso(alpha=1,max_iter=1000,tol=.01)
lassoreg.fit(Y_train,X_train)
pred=lassoreg.predict(y_test)
rmse2=mean_squared_error(x_test,pred)**0.5
print(rmse2)
```

# Ridge Regression

```python
from sklearn import linear_model
ridgereg=linear_model.Ridge(alpha=1,max_iter=1000,tol=.01)
ridgereg.fit(Y_train,X_train)
pred1=ridgereg.predict(y_test)
rmse3=mean_squared_error(x_test,pred1)**0.5
print(rmse3)
```

# KNN Algorithm

```python
from sklearn.neighbors import KNeighborsRegressor
knnreg=KNeighborsRegressor(n_neighbors=200)
knnreg.fit(Y_train,X_train)
pred2=knnreg.predict(y_test)
rmse4=mean_squared_error(x_test,pred2)**0.5
print(rmse4)
```

# SVR Model

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
sc1=StandardScaler()
sc2=StandardScaler()
y1=Y_train.copy()
y2=y_test.copy()
y1=sc1.fit_transform(y1)
y2=sc2.fit_transform(y2)
svr=SVR()
svr.fit(y1,X_train)
pred3=svr.predict(y2)
rmse5=mean_squared_error(x_test,pred3)**0.5
print(rmse5)
```

# Naive Bayes Algorithm

```python
from sklearn.linear_model import BayesianRidge
bayreg=BayesianRidge()
bayreg.fit(Y_train,X_train)
pred4=bayreg.predict(y_test)
rmse6=mean_squared_error(x_test,pred4)**0.5
print(rmse6)
```

# Random Forest Algorithm

```python
from sklearn.ensemble import RandomForestRegressor
rand=RandomForestRegressor()
rand.fit(Y_train,X_train)
pred5=rand.predict(y_test)
rmse7=mean_squared_error(x_test,pred5)**0.5
print(rmse7)
```

# LightGBM Model

```python
from lightgbm import LGBMRegressor
lgbm=LGBMRegressor()
lgbm.fit(Y_train,X_train)
pred6=lgbm.predict(y_test)
rmse8=mean_squared_error(x_test,pred6)**0.5
print(rmse8)
```
