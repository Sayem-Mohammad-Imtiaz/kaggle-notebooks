```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
#set up file paths
salaries_filepath = "../input/us-major-league-soccer-salaries/mls-salaries-2007.csv"

#read the file
salaries_data = pd.read_csv(salaries_filepath, index_col="position")

#confirm proper load
salaries_data.head()
```

```python
#create heatmap
plt.figure(figsize=(12,6))

sns.set_style("whitegrid")

sns.barplot(x=salaries_data.index, y=salaries_data['base_salary'])

plt.title("MLS Base Salary Per Position 2007")
plt.xlabel("Position")
plt.ylabel("Base Salary")
```
