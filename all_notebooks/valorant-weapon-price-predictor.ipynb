```python
df = pd.read_csv("../input/valorant-weapons-stats/valorant-weapons.csv")
df
```

```python
df = pd.get_dummies(df,columns=["type","wall-pen"])
```

```python
df
```

```python
X = df.iloc[:,2:].values
X
```

```python
y = df.iloc[:,1:2].values
y
```

```python
X.shape[1]
```

```python
y = tf.cast(y,tf.float32)
y
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.BatchNormalization(input_shape=(X.shape[1],)))
model.add(tf.keras.layers.Dense(100,activation="relu"))
model.add(tf.keras.layers.Dense(1))
model.add(tf.keras.layers.Lambda(lambda x : x * 1000))

model.summary()
```

```python
def loss(y_true,y_pred) :
    error = tf.math.reduce_euclidean_norm(tf.square(y_true - y_pred))
    return error
```

```python
optimizer = tf.keras.optimizers.Adam(0.01)
model.compile(optimizer=optimizer,loss=loss)
```

```python
model.fit(X,y,epochs=500,verbose=0)
```

```python
odin = model.predict([[12,100,95,38,32,95,38,32,77,31,26,1,0,0,0,0,0,1,0,0]]) # Odin
```

```python
print(f"Predicted Odin's price is {odin[0][0]}")
```
