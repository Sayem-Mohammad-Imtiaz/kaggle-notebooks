```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
spotify_data = pd.read_csv("../input/spotify-dataset-19212020-160k-tracks/data.csv", index_col="year")
```

```python
# show spotify_data content
spotify_data
```

```python
# show 5 lasts spotify_data content
spotify_data.tail()
```

```python
# relation between release date and speechiness of songs
# is the more years go by, the less songs have words
plt.figure(figsize = (50, 10))

plt.title("Relation between songs release date and speechiness")
sns.barplot(x = spotify_data.index, y = spotify_data['speechiness'])
plt.ylabel("Indice of speechiness")
```

**the graph says that there is no need to say that the more the years pass, the less the songs have less meaning (speechiness)**

```python
# Are songs become more explicit ?
plt.figure(figsize = (50, 10))
plt.title("Evolution of explicit songs")
sns.scatterplot(x = spotify_data.index, y = spotify_data['speechiness'], hue = spotify_data['explicit'])
```

**We can say that newer songs are more explicit**

```python
# Is the more a song is danceable, the more populare it is
plt.figure(figsize = (50, 10))

plt.title("Relation between songs popularity and danceable")
sns.barplot(x = spotify_data['danceability'], y = spotify_data['popularity'])
plt.ylabel("Relation danceability/popularity")
```

**The more a song is danceable, the more it is populate but it's not that so obvious**
