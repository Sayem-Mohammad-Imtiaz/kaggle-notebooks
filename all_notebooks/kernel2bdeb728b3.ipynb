```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df=dataset = pd.read_csv("../input/haberman.csv",header=None,
        names=['age', 'year', 'nodes', 'status'])
print(df.shape)
print(df.columns)
df
```

```python
df["status"].value_counts()
```

obsevation:The dataset is imbalanced. The number of patients surviving for 5 years or more is greater than (almost 3 times) the number of patients that died within 5 years.

```python
sns.set_style("whitegrid")
sns.FacetGrid(df,hue="status",height=3).map(plt.scatter,"age","year").add_legend()
plt.show()
```

NOTE: the two categoies are not easily separable using attribute year and age. 

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(df,hue="status",vars=[df.columns[0],df.columns[1],df.columns[2]],height=3)
plt.show()
```

observation: Pair plot between age year and auxiliary nodes do not convey much and hence are not linearly separable. However the scatter plot between year of operation and auxiliary node gives better separation than others.

```python
#sinlge attribute analysis
for i,attrbt in enumerate(list(df.columns)[:-1]):
    sns.FacetGrid(df,hue="status",height=4).map(sns.distplot,attrbt).add_legend()
    plt.show()
```

some sort of observation can be obtained by the number of auxiliary nodes the patient has.Hence we can analyse status 1 and status 2 separately only with respect to number of nodes.

```python
before5=df.loc[df["status"]==2]
after5=df.loc[df["status"]==1]
```

```python
counts,bin_edges=np.histogram(before5["nodes"],bins=10,density=True)
pdf=counts/(sum(counts))
print("PDF")
print(pdf)
print("BIN EDGES")
print(bin_edges)
cdf=np.cumsum(pdf)
print("CDF")
print(cdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.xlabel("Nodes")
```

Observation:we see that almost 97.53 % of the person who have died within 5 years have auxiliary nodes less than 26

```python
counts,bin_edges=np.histogram(after5["nodes"],bins=10,density=True)
pdf=counts/(sum(counts))
print("PDF")
print(pdf)
print("BIN EDGES")
print(bin_edges)
cdf=np.cumsum(pdf)
print("CDF")
print(cdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.xlabel("Nodes")
```

Observation: almost 83.48 % of the people who survived more than 5 years have auxiliary nodes less than 4

```python
print(np.percentile(after5['nodes'],(25,50,75)))
print(np.percentile(before5['nodes'],(25,50,75)))
fig,axes=plt.subplots(1,3,figsize=(15, 5))
for i,atrbt in enumerate(list(df.columns)[:-1]):
    sns.boxplot(x='status',y=atrbt,data=df,ax=axes[i])
plt.show()
```

```python
print(np.percentile(after5['nodes'],(25,50,75)))
print(np.percentile(before5['nodes'],(25,50,75)))
fig,axes=plt.subplots(1,3,figsize=(15, 5))
for i,atrbt in enumerate(list(df.columns)[:-1]):
    sns.violinplot(x='status',y=atrbt,data=df,ax=axes[i])
plt.show()
```

we can observe that:
