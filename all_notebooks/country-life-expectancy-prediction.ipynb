```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
%matplotlib inline
```

## **Frequency of each country**

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,10))
plt.title('Countries data')
plt.xlabel('Frequency')
plt.ylabel('Country')
df.Entity.value_counts().plot(kind='barh')
plt.show()
```

## **Visualizing Each Country's Life Epxpectancy** 

```python
countries = df['Entity'].unique()
years = np.arange(1800,2017)
for country in countries:
    data = df[['Life expectancy','Year']][df['Entity'] == country ]
    
    plt.figure(figsize=(7,7))
    plt.plot(data['Year'], data['Life expectancy'], color='b', linewidth=1)
    plt.title(f"{country}'s Life expectancy")
    plt.xlabel('year')
    plt.ylabel('Life Expectancy')
    plt.show()
    
    print('\n')
```

### **One Hot Encoding on country data**

```python
country = pd.get_dummies(data=df['Entity'],prefix='country') #drop_first=True ;
```

```python
country
```

```python
new_df = pd.concat([country,df.drop(['Entity'],axis=1)],axis=1)
```

```python
new_df
```

### **Scaling the years**

```python
new_df['Year'] = new_df['Year']/10000
```

```python
new_df['Year'].plot(kind='hist',rwidth=0.8)
plt.show()
```

## Splitting and Training data 

```python
x = new_df.drop(['Life expectancy'], axis=1)
y = new_df['Life expectancy']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

### **Predictions and Accuracy** 

```python
y_pred = model.predict(x_test)
```

```python
y_pred_train = model.predict(x_train)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
mean_squared_error(y_train, y_pred_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```
