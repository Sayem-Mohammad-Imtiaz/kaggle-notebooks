### Сводные таблицы (Pivot Tables)
![](https://habrastorage.org/getpro/habr/post_images/85d/1d3/37a/85d1d337ac6b5671d819764785decde9.gif)

1. Здесь мы используем метод sales.pivot_table или crosstab, сообщая pandas о том, что мы хотим создать сводную таблицу, основанную на датафрейме sales.
2. Аргумент index указывает на столбец, по которому мы хотим агрегировать данные.
3. Аргумент values указывает на то, какие значения мы собираемся агрегировать.
4. Аргумент aggfunc задаёт функцию, которую мы хотим использовать при обработке значений (тут ещё можно воспользоваться функциями mean, max, min и так далее).

```python
df.head()
```

```python
df['Total calls'] = df['Total day calls'] + df['Total eve calls'] + \
                  df['Total night calls'] + df['Total intl calls']
df['Total charge'] = df[['Total day charge', 'Total eve charge',
                         'Total night charge', 'Total intl charge']].sum(axis=1)
```

```python
vehicle = re.compile(r'\b(дорог|светофор|дорожног|автобус|трамва|троллейб|проезд|мост)')
electricity = re.compile(r'\b(электрич|свет|энерго)')
gas = re.compile(r'\b(газов|газ)')
infrastructure  = re.compile(r'\b(торговый центр|набережн|парк|детск[а-я]+ площад|двор|клумб|деревь)')
medicine = re.compile(r'\b(госпитал|больниц|аптек|клиник|поликлиник)')
housing = re.compile(r'\b(жилье|дома)')
water = re.compile(r'\b(водопров|вода|воду|водоснабж|водоотвед|трубопров|труб|гидро)')
tourism = re.compile(r'\b(турист|отель|отели)')
garbage = re.compile(r'\b(мусор|полигон|коммунальн[а-я]* отходы)')
warm = re.compile(r'\b(тепло)')
sport = re.compile(r'\b(спорт|стадион|каток|поле)')
education = re.compile(r'\b(школа|детский сад|садик|лицей|образован)')

branches = {
'Транспорт': vehicle,
'ЭС': electricity,
'Газоснабжение': gas,
'Инфраструктура': infrastructure,
'Медицина': medicine,
'ЖКХ': housing,
'ВиВ': water,
'Туризм': tourism,
'ТКО': garbage,
'Тепло': warm,
'Спорт': sport,
'Образование':education
}

def get_branch(text):
    text_branches = ''
    max_finds=0
    for branch, regular in branches.items():
        finds = len(regular.findall(text))
        if finds>max_finds:
            max_finds = finds
            text_branches = branch
            
    return text_branches
```

```python
news['branch'] = news['title'].apply(get_branch)
```
