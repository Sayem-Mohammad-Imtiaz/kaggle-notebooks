We perform KMeans clustering on this dataset.We have few errors and the algorithm is using feature to cluster the gender into two distinct group.As the kmeans clustering is used for unsupervised data, we don't get a very good result over here.

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
kmeans.fit(df.drop("Gender",axis=1))
```

```python
kmeans.cluster_centers_
```

```python
def converter(Gender):
    if Gender=="Female":
        return 1
    else:
        return 0
    
```

```python
df["cluster"]=df["Gender"].apply(converter)
```

```python
df.head()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(df["cluster"],kmeans.labels_))
print("\n")
print(classification_report(df["cluster"],kmeans.labels_))
```
