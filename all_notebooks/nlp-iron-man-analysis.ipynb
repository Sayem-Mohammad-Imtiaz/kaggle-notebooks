```python
import re
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
import string
import nltk
import warnings 
warnings.filterwarnings("ignore", category=DeprecationWarning)

%matplotlib inline
```

```python
df = pd.read_csv("../input/iron-man-synopsis/iron man synopsis.csv")
df.head()
```

```python
#Removing Punctuations, Numbers, and Special Characters

df['Synopsis'] = df['Synopsis'].str.replace("[^a-zA-Z]", " ")
df.head()
```

```python
#remove all the words having length 3 or less

df['Synopsis'] = df['Synopsis'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))

df.head()
```

```python
#Tokenization: Tokens are individual terms or words, and tokenization is the process of splitting a string of text into tokens.

tokenized_Ironman = df['Synopsis'].apply(lambda x: x.split())
tokenized_Ironman.head()
```

```python
#Stemming: Stemming is a rule-based process of stripping the suffixes (“ing”, “ly”, “es”, “s” etc) from a word.

from nltk.stem.porter import *
stemmer = PorterStemmer()

tokenized_Ironman = tokenized_Ironman.apply(lambda x: [stemmer.stem(i) for i in x]) # stemming
tokenized_Ironman.head()
```

```python
#join the tokens back together

for i in range(len(tokenized_Ironman)):
    tokenized_Ironman[i] = ' '.join(tokenized_Ironman[i])

df['Synopsis'] = tokenized_Ironman
```

```python
df.head()
```

```python
#Wordcloud: A wordcloud is a visualization wherein the most frequent words appear in large size and the less frequent words appear in smaller sizes.

all_words = ' '.join([text for text in df['Synopsis']])
from wordcloud import WordCloud
wordcloud = WordCloud(width=800, height=500, random_state=21, max_font_size=110).generate(all_words)

plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()
```

```python
#features
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer(max_features=200, stop_words='english')
# bag-of-words feature matrix
#create dictionary from corpus
feature_vector = count_vectorizer.fit(df.Synopsis)

#get feature names
features = feature_vector.get_feature_names()

#Transform the records into vectors
df_features = count_vectorizer.transform(df.Synopsis)

#Count the frequency of features
features_counts = np.sum(df_features.toarray(), axis = 0)
feature_counts = pd.DataFrame(dict(features = features, counts = features_counts))
```
