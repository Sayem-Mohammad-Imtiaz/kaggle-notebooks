```python
#importing libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#importing dataset
ds = pd.read_csv('../input/position-salary-dataset/Position_Salaries.csv')
ds.head()
```

```python
#since position column and level column are one of the same thing we'll take only level column as our independent variable
x = ds.iloc[:,1:-1].values
y = ds.iloc[:,-1].values
print(x)
print(y)
```

**SPLITTING THE DATASET**
* as only 10 rows are present in the dataset, no splitting will be done so that we can have more data to train our model
* splitting will decrease accuracy

**FEATURE SCALING**
* AS SUPPORT VECTOR REGRESSION HAVE AN IMPLICIT EQUATION FOR DEPENDENT VARIABLE AND INDEPENDENT VARIABLE, THEREFORE FEATURE SCALING IS NECESSARY IN SVR
* STANDARDIZATION WILL BE DONE

```python
#applying feature scaling
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x = sc_x.fit_transform(x)
sc_y = StandardScaler()
y = sc_y.fit_transform(y.reshape(-1,1))
print(x)
print(y)
```

```python
#training the model
from sklearn.svm import SVR
svr = SVR(kernel = 'rbf')
svr.fit(x,y.flatten())
```

**VISUALIZING PREDICTION RESULTS**

```python
plt.scatter(sc_x.inverse_transform(x),sc_y.inverse_transform(y),c = 'red')
plt.plot(sc_x.inverse_transform(x), sc_y.inverse_transform(svr.predict(x)))
```

```python
from sklearn.metrics import r2_score
r2_score(y,svr.predict(x))
```

ACCURACY CAME TO BE 75.1 %
