```python
X = df.iloc[:,0:2]
y = df['Purchased']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X = scaler.fit_transform(X)
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,random_state=1) # 70% training and 30% test
```

# **LINEAR REGRESSION**

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True);
```

```python
print(classification_report(y_test, y_predict))
```

# **GRADIENT BOOSTING CLASSIFIER**

```python
modelGB = GradientBoostingClassifier()
modelGB.fit(X_train, y_train)
```

```python
predictGB = modelGB.predict(X_test)
```

```python
cm = confusion_matrix(y_test, predictGB)
sns.heatmap(cm, annot=True);
```

```python
print('Gradient Boost Classifier : \n', classification_report(predictGB, y_test))
print('Gradient Boost Classifier Accuracy: ', accuracy_score(predictGB, y_test))
```
