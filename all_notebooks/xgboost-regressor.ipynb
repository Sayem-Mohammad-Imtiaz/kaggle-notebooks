```python
df.drop(["UmId","ProductId"],axis=1)
```

```python
df_dummies = pd.get_dummies(df[['Place', 'ProductName', 'UmName']],drop_first=True)
```

```python
df_dummies
```

```python
from sklearn import preprocessing

min_max_scaler = preprocessing.MinMaxScaler()




df["Scaled_year"]=min_max_scaler.fit_transform(df[["Year"]])

df["Scaled_year"]
```

```python

df["Scaled_month"]=min_max_scaler.fit_transform(df[["Month"]])

df["Scaled_month"]

df.head()
```

```python


df_main = pd.concat([df_dummies,df], axis=1)

df_main.head()
```

```python
df_main.drop(['Place','ProductName','UmName'],axis=1)
```

```python
df_main.columns
```

```python
df_new_main=df_main.drop(['Place','ProductId', 'ProductName', 'UmId', 'UmName', 'Month', 'Year'],axis=1)
```

```python
df_new_main.columns
```

```python
from sklearn.model_selection import cross_val_score, train_test_split

x, y = df_new_main.drop(['Price'],axis=1), df_new_main['Price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25)

```

```python
import xgboost as xg
import numpy as np
from sklearn.metrics import mean_squared_error as MSE
```

```python
# Instantiation
xgb_r = xg.XGBRegressor(objective ='reg:linear',
                  n_estimators = 10, seed = 42)
```

```python
xgb_r.fit(x_train, y_train)
  
# Predict the model
pred = xgb_r.predict(x_test)
  
# RMSE Computation
rmse = np.sqrt(MSE(y_test, pred))
print("RMSE : % f" %(rmse))
```

```python
xgb_r.score(x_train,y_train)
```

```python
xgb_r.score(x_test,y_test)
```
