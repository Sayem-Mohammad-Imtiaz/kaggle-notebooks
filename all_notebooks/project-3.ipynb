```python
%matplotlib inline

import math
import random
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import csv
from matplotlib import pyplot as plt
from typing import *
post = open("/kaggle/input/washingtonposthomicidedata/fatal-police-shootings-summary.csv")
data = []
for line in post:
    data.append(line.split())
data=data[1:]
```

    The Washington Post's database contains records of every fatal shooting in the United States by a police officer in the line of duty since Jan. 1, 2015.
    
    In 2015, The Post began tracking more than a dozen details about each killing — including the race of the deceased, the circumstances of the shooting, whether the person was armed and whether the person was experiencing a mental-health crisis — by culling local news reports, law enforcement websites and social media, and by monitoring independent databases such as Killed by Police and Fatal Encounters. The Post conducted additional reporting in many cases.
* RACE: race of victim (str)
* -W: White, non-Hispanic
* -B: Black, non-Hispanic
* -A: Asian
* -N: Native American
* -H: Hispanic
* -O: Other
* -None: unknown
* ARMED: if victim was armed (str)
* -Unarmed
* -Armed
* SIGNS OF MENTAL ILLNESS: True if the victim had a mental illness (Boolean)
* CITY: where the shooting took place (str)

```python
race_counts = {}
for row in data:
    race = row[0]
    if race not in race_counts:
        race_counts[race] = 0
    race_counts[race] += 1
```

```python
race_counts
```

```python
plt.bar(race_counts.keys(),race_counts.values())
plt.title ('Victims By Race')
plt.xlabel('Race')
plt.ylabel('# of Victims')
plt.show()
```

```python
def city_graph():
    with open('/kaggle/input/washingtonposthomicidedata/fatal-police-shootings-summary.csv', newline='') as data_file:
        reader = csv.reader(data_file, delimiter=' ', quotechar='|')
        cities = []
        data = {}
        incidents = {}
        for row in reader:
            if row[3] not in cities:
                cities.append(row[3])
            for city in cities:
                if row[3] == city:
                    incident = row[1]
                    if incident not in incidents:
                        incidents[incident] = 0
                        data[city] = incidents
                    incidents[incident] += 1
                    data[city] = incidents

    for city in data:
        plt.bar(data[city].keys(), data[city].values())
        plt.ylabel('Number of Incidents')
        plt.xlabel(city)
        plt.show()
#if _name_ = 'main'
city_graph()
```
