**Descrição**
Sistema de apoio a decisão com base em histórico de vendas de veículo.

**Objetivo:**
Uma empresa de revenda de automóveis necessita saber qual veículo da marca de wolksvagen adiquirir, com isso deseja rever seu histórico de vendas e verificar os modelos mais vendidos

```python
# Importando as bibliotecas
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import matplotlib.pyplot as plt
import seaborn as sns

# Impressão dos dados
data = pd.read_csv("../input/vwcars/vw.csv")
```

# Processo ETL

```python
# Verificando se há dados vazios
data.isnull().sum()
```

Não há dados nulos, e também não há necessidade de tratar os dados

```python
# Exibindo a base
data.describe()
```

# # Apoio a visualização dos dados
# Coorelação das variáveis

```python
plt.figure(figsize=(20, 20))
sns.heatmap(data.corr(), annot=True, linewidths=0.01, cmap='coolwarm', cbar=False)
plt.show()
```

Podemos verificar que a variável preço está coorelacionada com o valor, logo adiquirir veiculos mais novos, acarreta em veículos mais caros, e vender veículos mais novos gera maior receita.

# Modelo dos veículos mais vendidos

```python
sns.countplot(y = data["model"])
```

## Conclusão

O modelo de dados permitiu visualizar quais os modelos de carros mais consumidos no mercado, e assim ajudar no direcionamento da compra de um veículo para sua posterior revenda
