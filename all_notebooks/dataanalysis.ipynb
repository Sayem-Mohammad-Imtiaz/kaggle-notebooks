```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
penguins=pd.read_csv('../input/penguins/penguins.csv').dropna()
```

```python
penguins.head()
```

```python
g=sns.FacetGrid(penguins,col='island')
g.map(sns.histplot,"flipper_length_mm");
```

```python
g=sns.FacetGrid(penguins,col='island')
g.map_dataframe(sns.scatterplot,"flipper_length_mm",y='bill_length_mm');
```

```python
sns.catplot(x="flipper_length_mm",data=penguins,kind='box');
```

```python
sns.catplot(x="flipper_length_mm",data=penguins,kind='violin');
```

```python
sns.catplot(x="species",y='bill_length_mm',data=penguins,hue='island',kind='box');
```

```python
sns.catplot(x="species",y='bill_length_mm',data=penguins,hue='island',kind='violin',palette='winter',whis=1);
```

```python
sns.relplot(x="species",y='bill_length_mm',data=penguins,hue='island',palette='spring',kind='line')
```

```python
sns.relplot(x="bill_length_mm",y='species',data=penguins,hue='island',palette=['black','red','green'],s=200);
```
