```python
df['work_type'] = df['work_type'].apply(lambda x: 'Self-employed' if x=='Govt_job' or x=='Never_worked' or x=='Self-employed' else 'Private')
df['gender']=df['gender'].apply(lambda x:'Female'if x=='Other' or x=='Female' else 'Male' )
df['smoking_status']=df['smoking_status'].apply(lambda x:'never smoked'if x=='Unknown' or x=='never smoked' else 'smokes')

```

```python
df
```

```python
df['gender'] = df['gender'].apply(lambda x: 1 if x=='Male'else 0)
df['ever_married'] = df['ever_married'].apply(lambda x: 1 if x=='Yes'else 0)
df['work_type'] = df['work_type'].apply(lambda x: 1 if x=='Private'else 0)
df['Residence_type'] = df['Residence_type'].apply(lambda x: 1 if x=='Urban'else 0)
df['smoking_status'] = df['smoking_status'].apply(lambda x: 1 if x=='smokes'else 0)
```

```python
df
```

```python
for i in df.columns:
    print(i,df[i].unique())
```

```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(df.drop('stroke',axis=1), df['stroke'], test_size=0.3, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
model = DecisionTreeClassifier(max_depth=1)
model.fit(X_train,y_train)
pred = model.predict(X_test)
```

```python
pred = model.predict(X_train.values)
print("Training Performance")
print()
print(classification_report(y_train,pred))
confusion_matrix(y_train,pred)
```
