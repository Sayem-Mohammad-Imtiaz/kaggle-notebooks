```python
import sys
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
%matplotlib inline


data = pd.read_csv("../input/nasa-access-log-dataset-1995/data.csv")
data.rename( columns={'Unnamed: 0':'row_index'}, inplace=True )
data.head()

```

```python
import datetime

data["timestamp"] =list(map(lambda x : datetime.datetime.fromtimestamp(int(x)).strftime('%Y-%m-%d %H:%M:%S'), data['time']))
data=data.drop(columns='time')
data.head()
```

```python
data['timestamp'] = pd.to_datetime(data.timestamp) 
data.head()
```

```python
data["date"]=data['timestamp'].dt.strftime("%Y-%m-%d")
data["time"]=data['timestamp'].dt.strftime("%H:%M:%S")

data.head()
```

```python
data.groupby('response').size().plot(kind='bar')
plt.title("status of the response")
plt.xlabel("success")
plt.legend(['status of response'],loc="upper right")
plt.show()
```

```python
success = data[data.response == 200]
success.head()
```

```python
error = data[data.response > 200]
error.head()
```

```python
import sys
plt.plot(success.response,success.bytes,label = 'success')
plt.plot(error.response,error.bytes,label='error')

plt.title("success vs error")
plt.xlabel("success")
plt.ylabel("bytes")
plt.legend()
plt.show()
```

```python
plt.plot(success.date.iloc[1:50],success.response.iloc[1:50],label = 'success')

plt.title("success responses vs date")
plt.xlabel("Date")
plt.ylabel("success")
plt.legend()
plt.show()
```

```python
plt.plot(success.date.iloc[1:20],success.response.iloc[1:20],label = 'success')
plt.plot(error.date.iloc[1:20],error.response.iloc[1:20],label='error')

plt.title("response status vs date")
plt.xlabel("Date")
plt.ylabel("success")
plt.legend()
plt.show()
```
