```python
#step 5:Determine the Features and labels:
X = df.drop('Sales',axis=1)
y = df['Sales']
```

```python
# Step 6: Split the Dataset to Train & Test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)

#assigning 30 percents of data to test data for evaluating the model
```

```python
#Step 7:Train the Model
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train, y_train)
pd.DataFrame(model.coef_ , X.columns ,columns=['coefcient'])

#these codes returns a DataFaram of coeficents
```

```python
#step 8:predicting Test Data
y_pred= model.predict(X_test)
y_pred
```

```python
np.array(y_test)
```

we can compare these numbers together 

```python
#Step 9: Evaluating the Model
from sklearn import metrics
MAE=metrics.mean_absolute_error(y_test , y_pred)
MSE=metrics.mean_squared_error(y_test , y_pred)
RMSE=np.sqrt(MSE)
a=pd.DataFrame([MAE,MSE,RMSE], index=['MAE','MSE','RMSE'],columns=['Quantity'])
a
```

```python
#Step 10:Residuals:
test_residual=y_test-y_pred
sns.scatterplot(x=y_test,y=y_pred,color='green' ,s=200)
plt.ylabel('y_pred')
plt.xlabel('y_test')
plt.title('bias of y')
```

The plot shows the label following a linear model

```python
sns.scatterplot(x=y_test,y=test_residual,s=200)
plt.axhline(y=0,color='red',ls='--')
```

It shows that residuals have been selected accidently so the plot doesn't have a specific pattern

```python
#Step11: Saving and Loading the Model
from joblib import dump, load
final_model=LinearRegression()
final_model.fit(X,y)
dump(final_model, 'Sales.joblib')
```

```python
load_model=load('Sales.joblib')
#it loads the model
```
