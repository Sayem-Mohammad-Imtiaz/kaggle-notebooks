# Missing Values

```python
print('Train miss\n\n',df_train.isna().sum())
print('\n\nTest miss\n\n',df_test.isna().sum())
```

Find Missing Values

```python
total= np.sum(df_train.isna()).sort_values(ascending =False)
persent = np.sum(df_train.isna())/(df_train.isna()).count()
df_train_missdata =pd.concat([total ,persent],axis = 1 , keys = ['total','persent'] )
nan_train = df_train_missdata
print('nan_train\n\n',nan_train)
print('----------------------------------------------')
total= np.sum(df_test.isna()).sort_values(ascending =False)
persent = np.sum(df_test.isna())/(df_test.isna()).count()
df_test_missdata =pd.concat([total ,persent],axis = 1 , keys = ['total','persent'] )
nan_test = df_test_missdata
print('nan_test\n\n',nan_test)
```

Find Missing Values and insert NAN for train data

داده های میس رو پیدا کردم و با nan پر کردم

```python
def cat_exploration(column):
    return df_train[column].value_counts()

def cat_imputation(column, value):
    df_train.loc[df_train[column].isnull(),column] = value
```

```python
list1 =nan_train.index

for i in list1 :
    cat_exploration(i)
    cat_imputation(i,'NA') 
    
```
