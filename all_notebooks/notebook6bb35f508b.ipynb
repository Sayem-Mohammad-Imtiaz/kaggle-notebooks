```python
data = pd.read_csv('/kaggle/input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv')
                
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score,f1_score
```

```python
data.isnull().sum()
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
from sklearn.linear_model import LinearRegression
imputer = IterativeImputer(estimator = LinearRegression())
data['bmi'] = imputer.fit_transform(data[['bmi']])
```

```python
X = data.iloc[:,:-1]
y = data['stroke']
```

```python
data.isnull().sum().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=402)
```

```python
categorical_colums = data.select_dtypes(include='object').columns
column=make_column_transformer((OneHotEncoder(),categorical_colums),remainder='passthrough')
```

```python
sm = SMOTE()
rf = RandomForestClassifier(n_estimators =100, max_depth = 7, max_leaf_nodes= 10)
pipe = make_pipeline(column,sm,rf)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_train = pipe.predict(X_train)
y_pred_test = pipe.predict(X_test)
print('Accuracy_score of Train:', accuracy_score(y_train,y_pred_train))
print('Accuracy_score of Test:', accuracy_score(y_test,y_pred_test))
```
