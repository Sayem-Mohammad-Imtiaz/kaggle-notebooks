```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (32, 15)
plt.rcParams["font.size"] = 24

f = open("/kaggle/input/average-rainfall-and-temperature-in-uk20102019/avg_rainfalltemp in UK - Sheet1.csv", "r", encoding="utf-8")

tarr = [10, 18, 19, 20, 21, 22, 24, 25, 26, 27, 34, 35, 36, 37, 39, 40, 42, 43, 44, 45, 46, 47, 48, 50, 51, 52, 53, 54, 55, 56, 57, 58, 60, 61, 62, 63, 64, 65, 66, 69, 70, 71, 73, 75, 76, 78, 79, 81, 82, 83, 85, 86, 87, 89, 93, 94, 95, 96, 104, 108, 111, 114, 115, 116, 118, 121, 124, 126, 127, 130, 140, 141, 143, 147, 148, 153, 157, 158, 160, 162, 172, 175, 177, 180, 182, 190, 191, 193, 196, 200, 204, 208, 209, 220, 225, 228, 231, 262, 267, 277, 279, 293, 322, 325, 344, 661, 705, 721, 738, 754, 778, 815, 831, 905, 1043]
outs = [0]*len(tarr)

lines = f.readlines()

print(lines[0])

for i in range(1, len(lines)):
    line = lines[i].split(",")
    outs[tarr.index(int(str(line[3]).split(".")[0]))] += float(line[4])

plt.plot(tarr, outs, linewidth=1.618, marker="o", label="Temperature")
plt.xlabel("Avg rainfall(in mm)")
plt.title("Temperature dependence on showers.")
plt.ylabel("Avg temp(in centigrade)")
plt.grid(linestyle="--", linewidth=1.618)
plt.legend()
plt.show()
```
