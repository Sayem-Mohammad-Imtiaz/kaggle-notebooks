```python

import numpy as np
import pandas as pd 
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter
```

```python
data = pd.read_csv('../input/pollution-index-of-pm25/Pollutant.csv')
data.head()
```

creating a timestamp from year month day hour columns

```python
#returns datetime object given some fields.
def to_timestamp(year,month,day,hour):
    return datetime(year,month,day,hour)
```

```python
#apply the above function to the required columns
data['time'] = data[['year','month','day','hour']].apply(
    lambda t: to_timestamp(t['year'],t['month'],t['day'],t['hour']),
    axis=1)
#drop the redundancy
data = data.drop(['year','month','day','hour'],axis=1)
#time is the new index
data = data.set_index('time')
```

Here is a function which plots time series with a moving average

```python
def plotter(data,var,roll=100,measure = 'measure',plot_size = (20,10)):
    """ 
    A function which plots both the raw data and a rolling mean
    data = pandas dataframe
    var = variable name
    measure = unit of measure (default=measure)
    roll = period of moving average
    """
    fig, ax = plt.subplots(figsize=plot_size)
    ax.plot(data.index,data[var],alpha=0.3) #raw
    data[var].rolling(roll).mean().plot(color='blue') #smoothed data
    
    #fix date format on plot
    date_form = DateFormatter("%Y-%m-%d")
    ax.xaxis.set_major_formatter(date_form)
    ax.set_title(f'Plot of {var} and {roll} period moving average')
    ax.set_xlabel('Date')
    ax.set_ylabel(measure)
    
plotter(data,'TEMP',300, 'degrees')
```

```python
plotter(data,'DEWP',roll=200,plot_size=(10,10))
```

```python

totals_by_day = data.groupby([pd.Grouper(freq='d')]).sum()
#plot of daily totals of TEMP with a weekly average
plotter(totals_by_day,'TEMP',7,'degrees')
```
