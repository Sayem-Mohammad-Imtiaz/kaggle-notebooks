```python
# Importing the library 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import PowerTransformer
import scipy.stats as stats
import statsmodels.api as sm
from sklearn.metrics import r2_score,mean_squared_error
```

```python
# Loading the data 

df=pd.read_csv("/kaggle/input/yeh-concret-data/Concrete_Data_Yeh.csv")
df.head()
```

```python
# Checking th information 

df.info()
```

```python
# Checking for the missing values 

df.isnull().sum()
```

There are no missing values in the dataset. 

```python
# Checking the distibution of data

sns.pairplot(df,diag_kind="kde")
plt.show()
```

```python
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
df.skew()
```

```python
df1=df.drop("csMPa",axis=1)
```

```python
# Correcting the skewness

pt=PowerTransformer()
df3=pt.fit_transform(df1)
```

```python
df4=pd.DataFrame(df3,columns=['cement', 'slag', 'flyash', 'water', 'superplasticizer',
       'coarseaggregate', 'fineaggregate', 'age'])
df4.head()
```

```python
df4.skew()
```

```python
x=df4
y=df["csMPa"]
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
# Base OLS model

xc=sm.add_constant(x)
ols=sm.OLS(y,xc)
ols_mod=ols.fit()
ols_mod.summary()
```

```python
# Linear Model

lr=LinearRegression()
lr.fit(xtrain,ytrain)
```

```python
print("Training Score :", lr.score(xtrain,ytrain))
print("Testing Score :", lr.score(xtest,ytest))
```

```python
# Ridge model

from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import GridSearchCV


rid=Ridge()
param={'alpha':[0.000001,0.00001,0.0001,0.01,0.1,0.5,1,3,5,10,15,100]}
grid=GridSearchCV(rid,param_grid=param,cv=5,scoring="r2")
hyp_rid=grid.fit(xtrain,ytrain)
hyp_rid.best_params_
```

```python
rid5=Ridge(alpha=5)
rid5.fit(xtrain,ytrain)

print("Training Score :", rid5.score(xtrain,ytrain))
print("Testing Score :", rid5.score(xtest,ytest))
```

```python
# lasso Model

las=Lasso()
param={'alpha':[0.000001,0.00001,0.0001,0.01,0.1,0.5,1,3,5,10,15,100]}
grid=GridSearchCV(rid,param_grid=param,cv=5,scoring="r2")
hyp_las=grid.fit(xtrain,ytrain)
hyp_las.best_params_
```

```python
las5=Lasso(alpha=5)
las5.fit(xtrain,ytrain)

print("Training Score :", las5.score(xtrain,ytrain))
print("Testing Score :", las5.score(xtest,ytest))
```

```python
# XGB Model


from xgboost import XGBRegressor

xgb=XGBRegressor()
xgb.fit(xtrain,ytrain)


print("Training Score :", xgb.score(xtrain,ytrain))
print("Testing Score :", xgb.score(xtest,ytest))
```
