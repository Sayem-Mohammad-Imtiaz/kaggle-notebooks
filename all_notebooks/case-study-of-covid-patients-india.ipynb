```python
# TO COMPARE LINE GRAPHS OF CONFIRMED AND CURED PATIENTS

import pandas as pd
df=pd.read_csv('../input/covid-cases-in-india/covid_data.csv')

df1=df.groupby(['State_UnionTerritory']).mean()['Confirmed']
df2=df.groupby(['State_UnionTerritory']).mean()['Cured']

sr1=pd.Series(df1)      
sr2=pd.Series(df2)

record=pd.DataFrame({'Confirmed':sr1,'Cured':sr2})
record.plot()
```

```python
# TO COMPARE LINE GRAPHS OF CONFIRMED AND DEAD PATIENTS

import pandas as pd
df=pd.read_csv('../input/covid-cases-in-india/covid_data.csv')

df1=df.groupby(['State_UnionTerritory']).mean()['Confirmed']
df2=df.groupby(['State_UnionTerritory']).mean()['Deaths']

sr1=pd.Series(df1)      
sr2=pd.Series(df2)

record=pd.DataFrame({'Confirmed':sr1,'Deaths':sr2})
record.plot()
```

```python
# TO COMPARE LINE GRAPHS OF CONFIRMED,CURED AND DEAD PATIENTS

import pandas as pd
df=pd.read_csv('../input/covid-cases-in-india/covid_data.csv')

df1=df.groupby(['State_UnionTerritory']).mean()['Confirmed']
df2=df.groupby(['State_UnionTerritory']).mean()['Cured']
df3=df.groupby(['State_UnionTerritory']).mean()['Deaths']

sr1=pd.Series(df1)      
sr2=pd.Series(df2)
sr3=pd.Series(df3)

record=pd.DataFrame({'Confirmed':sr1,'Cured':sr2,'Deaths':sr3})
record.plot()
```
