```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import pandasql as ps
```

```python
df = pd.read_csv("/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv")
df.head()
```

```python
#### Checking the null value and the aggregation on it 

df.isnull().sum()
```

```python
####### Replacing the Null value with 0 as the Vaccination is started recently there might be a probability of geting missing data here 

df.fillna(value = 0,inplace = True)
```

```python
######## Checking the shape of the data 

df.shape
```

```python
### Checking the Unique country along with the fetaures name associated with the dataframe
countries = df['country'].unique()
print(countries)
print(len(countries))
df.head()
```

```python
### We can check the total number of vaccines 

no_of_tot_vac = sum(df['total_vaccinations'])
no_of_tot_vac
```

```python
### total Vaccine by country and checking the 10 countries to be plot 

import pandasql as ps

country_by_vacc = ps.sqldf("""
 select a.iso_code , sum(a.total_vaccinations) as tot_vac ,a.vaccines  from df as a 
     group by a.iso_code 
     limit(10)
""")

country_by_vacc.head()
plt.plot(country_by_vacc.iso_code,country_by_vacc.tot_vac,linestyle= 'dashed')
```

```python
### Aggregating the data by country 
##### Answer of Task 1 
#### Selecting only the numerical features here 
col_index = ['country', 'iso_code', 'date', 'total_vaccinations',
       'people_vaccinated', 'people_fully_vaccinated',
       'daily_vaccinations_raw', 'daily_vaccinations',
       'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred',
       'people_fully_vaccinated_per_hundred', 'daily_vaccinations_per_million']
df_1 =  df[col_index]
df_1.head()

df_1_agg = df_1.groupby('country').sum()
df_1_agg
```

```python
df_1_agg.sort_values(by = 'total_vaccinations',ascending = False,inplace = True)
df_1_agg
```

```python
####### Task 2 Countries with highest number of total vaccinations


total_vac_sorted  = df_1_agg[0:10].sort_values(by = 'total_vaccinations',ascending = False) 


plt.style.use('ggplot')
total_vac_sorted.plot.barh()
plt.title('top 10 countries')
plt.ylabel('country')
plt.xlabel('total_vaccinations')
plt.show()
```

```python
###Task3: What country is vaccinated a larger percent from its population? 

##### Mean is the best measurement while we are calculating against the population here total vaccine canot be a god measurement.
people_vacc_df = df[['country', 'total_vaccinations_per_hundred']]
people_vacc_df.head()
people_vacc_grouped = people_vacc_df.groupby("country").mean()
people_vacc_grouped.head()
```

```python
##Sorting the data based on total_vaccinations_per_hundred from higher to lower
people_vacc_grouped.sort_values(by="total_vaccinations_per_hundred", ascending=False, inplace=True)
people_vacc_grouped.head()
```

```python
people_vacc_grouped_1 = people_vacc_grouped[0:10].sort_values(by="total_vaccinations_per_hundred")
plt.style.use('ggplot')
people_vacc_grouped_1.plot.barh(color = 'g')

plt.ylabel('country')
plt.xlabel('total_vaccinations_per_hundred')
plt.show()
```
