## Replace nulls
- Replace null values with median of column Lane.

```python
df["Lane"]=df["Lane"].fillna(df["Lane"].median())
```

```python
df.isnull().sum()
```

- Raplacing nulls in column Rank by value 33.

```python
df["Rank"].unique()
```

```python
df["Rank"]=df["Rank"].fillna(33)
```

```python
df["Rank"]=df["Rank"].replace("ï¿½",33)
```

## Visualization

```python
plt.subplot(1,4,1)
plt.subplots_adjust(left=0,right=2,bottom=1,top=2,wspace=0.2,hspace=0.4)
label_age=['Olympics', 'World Championship']
plt.title("Where")
plt.pie(df.groupby(df["Where"]).size(),labels=label_age)

plt.subplot(1,4,2)
label_gen=[2021, 2016, 2015, 2019]
plt.title("Year")
plt.pie(df.groupby(df["Year"]).size(),labels=label_gen)

plt.subplot(1,4,3)
label_ed=['Run', 'Hurdles', 'Steeplechase']
plt.title("Event")
plt.pie(df.groupby(df["Event"]).size(),labels=label_ed)

plt.subplot(1,4,4)
label_lw=['M', 'F']
plt.title("Sex")
plt.pie(df.groupby(df["Sex"]).size(),labels=label_lw)
plt.show()
```

```python
df["Rank"]=df["Rank"].astype("int")
```

```python
plt.subplot(1,2,1)
plt.subplots_adjust(left=0,right=2,bottom=1,top=2,wspace=0.2,hspace=0.4)
plt.xticks(rotation = "vertical")
plt.title("Event vs Distance")
sb.barplot(x="Event" ,y="Distance", data=df)

plt.subplot(1,2,2)
plt.title("Year")
sb.countplot(x="Year", data=df)
plt.show()

```
