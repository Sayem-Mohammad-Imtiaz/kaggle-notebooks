```python

def summlist(x,y,z):
    summ = []
    for i in range(0,len(x)):
        k = x[i]+y[i]+z[i]
        summ.append(k)
    return summ

math = [student*2/10 for student in df["math score"]]
read = [student*4/10 for student in df["reading score"]]
write = [student*4/10 for student in df["writing score"]]


df["verbal ability"] = summlist(math,read,write)
```

Creating function which can summ lists and adding new attribute to our data frame

```python
df.corr()
```

```python
average = sum(df["verbal ability"])/len(df["verbal ability"])

df["verbal succes"] = ["Great" if i>average else "Unsufficient" for i in df["verbal ability"]]
```
