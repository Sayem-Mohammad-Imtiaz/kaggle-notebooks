```python
dataset=[]
datapath=[]
dataid=[]

for im in os.listdir(directory):
    path=os.path.join(directory,im)
    image=load_img(path, grayscale=False, color_mode='rgb', target_size=(100,100))
    image=img_to_array(image)
    image=image/255.0
    idn=im[0:-4]
    dataset.append([image])
    datapath.append(path)
    dataid.append(idn)
```

```python
testset=[]
testpath=[]
testid=[]

for im in os.listdir(directory2):
    path=os.path.join(directory2,im)    
    image=load_img(path, grayscale=False, color_mode='rgb', target_size=(100,100))
    image=img_to_array(image)
    image=image/255.0
    idn=im[0:-4]
    testset.append([image])
    testpath.append(path)
    testid.append(idn)
    
```

# Yolo Data Viewing


```python
def draw_bbox2(num0):
    
    img0=datapath[num0]
    id0=dataid[num0]
    df=train[train['image_id']==id0]
    im = Image.open(img0)
    W,H = im.size
    _ = plt.figure(figsize = (15,20))
    _ = plt.axis('on')
    _ = plt.imshow(mpimg.imread(img0))
    ax = plt.gca()
    ax.text(W*0.02,H*0.04,f'{id0}',fontsize=20,color='yellow')

    #for i in range(len(df)):    # all
    #for i in [0,3,4]:           # select some
    for i in range(5):           # limited range
        dfi=df[i:i+1]
        name=dfi.iloc[0,1]
        W0=dfi.iloc[0,8]
        H0=dfi.iloc[0,9]
        x0=dfi.iloc[0,4]/W0
        y0=dfi.iloc[0,5]/H0
        x1=dfi.iloc[0,6]/W0
        y1=dfi.iloc[0,7]/H0
        w0=x1-x0
        h0=y1-y0
        x=x0*W
        y=y0*H
        w=w0*W
        h=h0*H
        rect = patches.Rectangle((x,y),w,h,linewidth=2,edgecolor='yellow',fill = False)
        ax.text(x,y-10,f'{name}',fontsize=12,color='yellow')
        ax.add_patch(rect)
        plt.show
```
