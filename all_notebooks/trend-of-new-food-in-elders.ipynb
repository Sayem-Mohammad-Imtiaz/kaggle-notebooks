* The task at hand was to check if people tend to be more willing to try new foods as they get old based on the data set provided.
* I have taken the approach to compare the frequency of willingness of certain food with ages of the individuals. 
* At the end you get bar charts as visual discriptions showing the split in people selection of food. I can be seen that for the given data set, the "elderly' tend not to try on new things as they get old.

```python
from matplotlib import pyplot as plt
import collections

def ploting_graph (x,y,xlabel,ylabel,title,kind,true_count):

    fig = plt.figure(figsize=(13, 7))
    plt.grid(True)
    plt.xlabel (xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    if len(x) != 0:
        plt.xticks(np.arange(min(x), max(x)+2, 2))
    if kind == 'bar':
        plt.bar(x,y,width=.8)
        plt.legend([true_count])
    elif kind == 'plot':
        plt.plot(x,y)
    else:
        print ('type : bar or plot ')

    return plt.show()

df = pd.read_csv('/kaggle/input/have-you-ever-eaten-these-foods/responses.csv')


headers = df.columns # Getting Headers from File
No_R = df.index.to_numpy()

for heads in headers:
    a = []
    true_count = 0
    if (heads == 'Age in years') or (heads =='Gender') or (heads == 'Race') or (heads =='Home location'):
        continue

    for ind in No_R:
        if str(df[heads][ind]) == ("True"):
            true_count +=1
            a.append(df["Age in years"][ind])
        
    count_ = collections.Counter(a)
    _ages = list(count_.keys())   
    _freq = list(count_.values())
    ploting_graph(_ages,_freq,'Ages','Frequency',f'People Who Can Eat {heads}','bar',true_count)
```
