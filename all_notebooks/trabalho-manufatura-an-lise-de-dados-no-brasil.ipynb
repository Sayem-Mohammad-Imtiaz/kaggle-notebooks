```python
dados['mdsal'] = dados["('P16', 'salary_range')"].fillna('$ 0/').apply(lambda x: #get the mean in ranges
                    int( #transform all in int in the end
                    (int(str(x)[str(x).rfind(' ')+1:str(x).rfind('/')].replace('.','')) # Get max in range 
                    +
                    int(str(x)[str(x).find('$')+2:str(x).find('/')].replace('.','')) #Get min
                    )/2)) #divide by 2
```

```python
dados.groupby("('P10', 'job_situation')")['mdsal'].mean().nlargest(6).plot.barh(color='red')
```

Homens Ganham mais que mulheres ?

```python
dados.groupby("('P2', 'gender')")['mdsal'].mean()
```

```python
dados.groupby("('P2', 'gender')")['mdsal'].mean().plot.bar(color=['#f5005a', 'blue'])
```

```python
col_ad = "('P17', 'time_experience_data_science')"
dados.loc[dados[col_ad] == 'Não tenho experiência na área de dados', col_ad] = 'Sem experiência'

men = dados[dados.loc[:, "('P2', 'gender')"] == 'Masculino'] #Linhas com sexo Masculino
men_state_mean = men.groupby("('P5', 'living_state')")['mdsal'].mean() #Média Masculina por região

woman = dados[dados.loc[:, "('P2', 'gender')"] == 'Feminino'] #Linhas com sexo Feminino
woman_state_mean = woman.groupby("('P5', 'living_state')")['mdsal'].mean() #Média Feminina por região

menxp = men.groupby("('P17', 'time_experience_data_science')")['mdsal'].mean()
womanxp = woman.groupby("('P17', 'time_experience_data_science')")['mdsal'].mean()
```

```python
plt.barh(men_state_mean.index, men_state_mean.values, color='blue')
plt.barh(woman_state_mean.index, woman_state_mean.values, color='#f5005a')

plt.xlabel("Média Salarial em R$")
plt.ylabel("Estados")

plt.title("Comparativo entre o salário de Homens e Mulheres por Região")

plt.legend(['Homens', 'Mulheres'])
```

```python
plt.barh(menxp.index, menxp.values, color='blue')
plt.barh(womanxp.index, womanxp.values, color='#f5005a')

plt.xlabel("Média Salarial em R$")
plt.ylabel("Experiência na área de dados")
plt.legend(['Homens', 'Mulheres'])

plt.title("Comparativo entre o salário de Homens e Mulheres com mesma experiência")
```

```python
sal_convertido = {
    'de R$ 1.001/mês a R$ 2.000/mês': 1500., 
    'de R$ 2.001/mês a R$ 3000/mês': 2500.,
    'de R$ 4.001/mês a R$ 6.000/mês': 5000., 
    'de R$ 6.001/mês a R$ 8.000/mês': 7000.,
    'de R$ 3.001/mês a R$ 4.000/mês': 3500.,
    'de R$ 8.001/mês a R$ 12.000/mês': 10000.,
    'de R$ 12.001/mês a R$ 16.000/mês': 14000.,
    'Menos de R$ 1.000/mês': 1000.,
    'de R$ 16.001/mês a R$ 20.000/mês': 18000.,
    'de R$ 20.001/mês a R$ 25.000/mês': 22500., 
    'Acima de R$ 25.001/mês': 25000.
}

dados['salario'] = dados["('P16', 'salary_range')"].map(sal_convertido)
```
