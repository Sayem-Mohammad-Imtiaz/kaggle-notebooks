**Bossa Nova is a laid-back song that is often found in cafes.** Bossa Nova is a genre of music that originated in Brazil and represents the jazz trend of the 1960s. Bossa Nova was created in the late 1950s by middle-class students and street musicians living in Rio de Janeiro. At Nara Liang's house overlooking Copacabana Beach, young and stylish musicians such as Antonio Carlos Zorbing, Vinicius g. Morais, Carlos Lira and Roberto Menescal took turns playing guitar and boasting new songs every week. In a free atmosphere, the first bossa nova song, Chega de Saudade, was born in 1958.

Bossa nova was first created in Brazil, developed and brought to the United States.

At that time, bossa nova was not just popular music, but a symbol of Brazil's modernization. **Bossa nova musicians pursued sophistication.** I tried to hide emotions such as love, passion, sadness, and longing in restrained lyrics and melodies without revealing them as they are. Such an attempt went beyond music and became a lifestyle symbolizing the youth and culture of middle-class Brazilian youth.

However, despite this success, as the military dictatorship began in Brazil after the mid-1960s, bossa nova gradually began to decline. **Because bossa nova, with its lyrical and hedonistic atmosphere, was not suitable for conveying a message of resistance to the dictatorship**, singers centered on Nara Liang pioneered a new music genre called MPB (Musica Populeira Brasileira, Brazilian popular music). started betting.

In Brazil, bossa nova has **become a cultured music mainly listened to by people of the middle class and above**, and although it is rather old, it is loved as a genre that symbolizes the country.


```python
df = df.sample(100)
```

```python
translator = Translator()

df['song_lyrics_ENG'] = df['song_lyrics'].apply(translator.translate, src='auto', dest='en').apply(getattr, args=('text',))
```

Translate 5000 lyrics takes some time. In my case, it takes about 1 hour to complete without using accelerator.
To make it short, I would like to select sample 100 songs randomly.

```python
df
```

```python

text_array = df['song_lyrics_ENG'].values
text_string = "".join(text_array)
text = text_string.replace("\n", " ")

IMG = Image.open(requests.get("https://www.pngarts.com/files/10/Coco-Guitar-Clipart-Download-Transparent-PNG-Image.png", stream=True).raw)
IMG_mask = Image.new("RGB", IMG.size, (255,255,255))
IMG_mask.paste(IMG,IMG)
IMG_mask = np.array(IMG_mask)

stopwords = set(STOPWORDS)
stopwords.add("WHEN_CASE_STOPWORDS_NEED")

wordcloud = WordCloud(width = 1200,
                      height = 1300,
                      max_words = 150,
                      stopwords = stopwords,
                      mask=IMG_mask,
                      background_color = "black",
                      contour_width=5,
                      contour_color='steelblue',
                      max_font_size = 200,
                      min_word_length= 2,
                      random_state=40
                     ).generate(text)

fig = plt.figure(figsize=(12,12))
plt.imshow(wordcloud)
plt.axis('off')
```
