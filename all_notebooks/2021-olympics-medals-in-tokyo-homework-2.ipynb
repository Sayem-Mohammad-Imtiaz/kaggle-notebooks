```python
# lambda function example
gpa = lambda x,y : (x+y)/2 # x is the average of the first term and y is the average of the second term
print(gpa(2.44,2.83))

# another example
eighteen_percent_of_any_number = lambda x : (x*18)/100
print(eighteen_percent_of_any_number(75))
```

```python
# anonymous function example, they are like lambda func. but they can take more than one argument
# map(func, seq): applies a function to all the items in a list
numbers = [33,66,99]
y = map(lambda x:(x*18)/100, numbers) # returns the eighteen percent of the numbers 
print(list(y))
```

```python
# nested function example
def multiplication():
    def summation():
        a = 963946208
        b = 2376477
        c = a + b
        return c
    return summation()* 3
print(multiplication())
```

```python
# default arguments: function arguments can have default values
def greeting(name, msg = 'Welcome'):
    """this function prints a greeting message specific to the given name if the message is not defined"""
    print(name,',', msg,'!')
greeting("rachellen")    
```

When you are defining a function with a default argument, make sure that the non-default argument comes first and the default argument comes later.

```python
# flexible arguments
# *args
def f(*args):
    for i in args:
        print(i,)
f(9,8,7)

# **kwargs
def f(**kwargs):
    for key, value in kwargs.items():
        print(key,':',value)
f(one = 'eins', two = 'zwei', three = 'drei')        
     
```

```python
# example of list comprehension
list1 = [7,5,3]
list2 = [i//2 for i in list1]
print(list2)
```
