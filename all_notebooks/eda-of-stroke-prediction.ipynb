```python
df_new=df.drop(['gender', 'ever_married',"work_type","Residence_type","smoking_status"], axis = 1)
```

```python
df_new
```

```python
df_X = df_new
```

```python
df_X_train = df_X[:-100]
df_X_test = df_X[-100:]
```

```python
df_y_train = df.avg_glucose_level[:-100]
df_y_test = df.avg_glucose_level[-100:]
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(df_X_train, df_y_train)
```

```python
df_y_predicted = model.predict(df_X_test)
print(df_y_predicted)
```

```python
print("Mean square error is: ",mean_squared_error(df_y_test, df_y_predicted))
```

```python
print("Weights: ", model.coef_)
print("Intercept: ", model.intercept_)
```

```python
df_X_test=np.arange(0,len(df_X_test),1)
```
