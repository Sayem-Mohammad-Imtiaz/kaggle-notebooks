```python
import pandas as pd
data = pd.read_csv("../input/college-data/data.csv")
```

```python
data.head()
```

```python
data[["apps", "accept", "enroll", "top10perc", "top25perc", "outstate", "phd", "grad_rate", "room_board"]].corr()
```

```python
from scipy import stats
pearson_coef, p_value = stats.pearsonr(data["outstate"], data["room_board"])

print(pearson_coef)

print(p_value)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x=data["outstate"], y=data["room_board"], marker= "P", color = "peachpuff", line_kws={'color':'chocolate'})
plt.show()
```

This graph shows that, on average, when there are more out of state students, the cost of rooms are more expensive. 

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
#df = sns.load_dataset('data')
 
# Make boxplot for one group only
sns.boxplot(x=data["outstate"], color="mistyrose", linewidth = 5)
plt.show()
```

This box plot shows the average number of out of state students per university. On average, there are 10,000 out of state students, which does not include international students, in universities. 

```python
# library & dataset
import seaborn as sns
import matplotlib.pyplot as plt
#df = sns.load_dataset('data')
 
# Basic 2D density plot
sns.set_style("white")
#sns.plt.show()
 
# Custom it with the same argument as 1D density plot
#sns.kdeplot(data, cmap="Reds", shade=True, bw=.15)
 
# Some features are characteristic of 2D: color palette and wether or not color the lowest range
sns.kdeplot(data["outstate"], data["room_board"], cmap="Blues", shade=True, shade_lowest=True)
plt.show()
```

This 2D density plot shows that the average amount of out of state students in universities, 10,000 students, pay approximately $4,000 for rooms, but these values differ as the number of students increases or decreases.
