```python
import matplotlib.pyplot as plt
%matplotlib inline
cdict = {0: 'red', 1: 'blue', 2: 'green',3:"lightblue",4:"orange"}
fig, ax = plt.subplots()
group = gmm.fit_predict(embeddingTest)
scatter_x = embeddingTest.to_numpy()[:,0]
scatter_y = embeddingTest.to_numpy()[:,1]

for g in np.unique(group):
    ix = np.where(group == g)
    ax.scatter(scatter_x[ix], scatter_y[ix], c = cdict[g], label = g, s = 2)
ax.legend()
plt.show()
```
