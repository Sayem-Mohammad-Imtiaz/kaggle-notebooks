```python
fig, ax = plt.subplots(43, 4, figsize=(15, 70))
[axi.set_axis_off() for axi in ax.ravel()]
i = 1
    
for cmap in colormap:

    ax = fig.add_subplot(41,4,i)
    ax.imshow(WordCloud(stopwords=STOPWORDS, colormap=cmap).generate(word_string))
    ax.title.set_text(cmap)
    plt.axis('off')
    i = i + 1

plt.show()
```
