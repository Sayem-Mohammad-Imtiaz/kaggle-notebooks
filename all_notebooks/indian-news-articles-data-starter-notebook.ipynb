# Data loading

```python
news = pd.read_csv('/kaggle/input/indian-news-articles/historic_articles.csv')
news.head()
```

```python
news.info()
```

# News source and categories

```python
print("Sources")
print(news['source'].value_counts())
```

```python
print("Category")
news.groupby(['source','category']).count()['link']
```

# Top Authors in each source/group

```python
print('firstpost/entertainment')
print(news[(news['source']=='firstpost') & (news['category']=='entertainment')].groupby(['author']).count()['link'].sort_values(ascending=False).head(3))
```

```python
s = news['author'].groupby([news['source'],news['category']]).value_counts()
s.groupby(level=[0,1]).head(3)
```
