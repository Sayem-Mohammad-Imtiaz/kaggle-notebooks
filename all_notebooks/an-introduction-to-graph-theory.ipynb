<h3 style="background-color:orange;font-family:newtimeroman;font-size:200%;text-align:center;border-radius: 15px 50px;">What Is A Graph?</h3>

<p style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; text-align: center;"><span data-preserver-spaces="true" style='color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; font-family: "Times New Roman", Times, serif; font-size: 24px;'>&nbsp;</span></p>
<p style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><em style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">Graphs</span></em><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;"> are mathematical structures used to model pairwise relations between objects. A graph in this context is made up of&nbsp;</span><em style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">vertices</span></em><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">&nbsp;(also called&nbsp;</span><em style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">nodes</span></em><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">&nbsp;or&nbsp;</span><em style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">points</span></em><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">) connected by&nbsp;</span><em style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">edges</span></em><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">&nbsp;(also called&nbsp;</span><em style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">links</span></em><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">&nbsp;or&nbsp;</span><em style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">lines</span></em></span></span><span data-preserver-spaces="true" style='color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; font-family: "Times New Roman", Times, serif; font-size: 24px;'>). <h3 style="background-color:orange;font-family:newtimeroman;font-size:200%;text-align:center;border-radius: 15px 50px;">Formal Defenition</h3>

<p style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; text-align: center;"><span data-preserver-spaces="true" style='color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; font-family: "Times New Roman", Times, serif; font-size: 24px;'>Now that we have a rough understanding of what is a graph, let&apos;s look at a formal definition of a graph;</span></p>
<p style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><br></span></span></p>
<p style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">Let $V$ be a finite set and $E\subseteq V \times V$such that for every $V_i,V_j \in E$ $|i\neq j$</span></span></span></p>
<ul style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">$G = (V,E)$ is a graph with $V$ vertices and $E$ edges.</span></span></span></li>
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">Every $v \in V$ is called a &quot;Vertex&quot; or a &quot;Node&quot; in G.</span></span></span></li>
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span data-preserver-spaces="true" style='color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; font-family: "Times New Roman", Times, serif; font-size: 24px;'>Every $(x,y) \in E$ is called an &quot;Edge&quot; in G</span></li>
</ul>

<a id="1.3"></a>
<h3 style="background-color:orange;font-family:newtimeroman;font-size:200%;text-align:center;border-radius: 15px 50px;">Edge Degrees and Euler's Formula</h3>

<p style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; text-align: center;"><span data-preserver-spaces="true" style='color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; font-family: "Times New Roman", Times, serif; font-size: 24px;'>Let $G(V, E)$ be an undirected graph and v in V.</span></p>
<p style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><br></span></span></p>
<ul style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">The neighborhood of $v$ (meaning all the nodes that can be reached from $v$) is denoted by $N(v)$, and $N(v)$ is a set containing all vertices. ( in the above example of graph $C_4$ each node has 2 neighbors because, from each vertex, you can reach only 2 other vertices)</span></span></span></li>
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">The&nbsp;</span><em style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">degree of a&nbsp;</span></em><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">vertex in $G$ is denoted as $deg(v)$ and is equal to the size of the vertex&apos;s neighborhood; $deg(v) = |N(v)|$</span></span></span></li>
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">If $deg(v) = 1$ than $v$ is called a &quot;Leaf&quot;</span></span></span></li>
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">if $deg(v) = \alpha$ for every $v \in V$ than G is called $\alpha$-regular. (In the above example of graph $C_4$ the graph is 2-regular because, for every vertex in $G$, the vertex&apos;s degree is 2.)</span></span></span></li>
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">The maximum degree of $G$ is denoted by $\Delta(G)$ and is defined as ; $\Delta(G) = max\{deg(v):v \in V\}$</span></span></span></li>
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">The minimum degree of $G$ is denoted by $\delta(G)$ and is defined as ; $\delta(g) = min\{deg(v):v \in V\}$</span></span></span></li>
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">$\Delta(G) = \delta(G)  \longleftrightarrow  G = \alpha$-regular</span></span></span></li>
</ul>
<p style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><br></span></span></p>
<p style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">Eulers Formula:</span></span></span></p>
<ul style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">For every undirected graph $G(V, E)$, the sum of all node degrees is equal the two times the number of edges; $ \sum\nolimits_{v \in  V}^{} deg(v) = 2|E|$</span></span></span></li>
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">If $G(V,E)$ is $\alpha$-regular than $ 2|E| = \sum\nolimits_{v \in  V}^{} deg(v) = \alpha * |V|$</span></span></span></li>
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span data-preserver-spaces="true" style='color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; font-family: "Times New Roman", Times, serif; font-size: 24px;'>$G(V, E)$ is a valid undirected graph and exists only if there is an even number of vertices with odd degrees.</span></li>
</ul>

```python
fig = plt.figure(figsize=(7,7))
plt.title(r'Graph : $\overline{P_{3} \cup 2K_1}$',fontsize=19,fontweight='bold')
plt.imshow(plt.imread('/kaggle/input/390-images-of-popular-graph-theory-graphs/Graph Images/g_co-P3U2K1.gif'))
plt.text(3,6,r'$V_1$',color='red',fontsize=23)
plt.text(53,6,r'$V_2$',color='red',fontsize=23)
plt.text(3,56,r'$V_3$',color='red',fontsize=23)
plt.text(53,56,r'$V_4$',color='red',fontsize=23)
plt.text(93,31,r'$V_5$',color='red',fontsize=23)

fig.text(.05,-.025,'$\Delta = 4$ | $\delta =2$  | $N({V_1}) =\{V_2,V_3,V_4 \}$',color='red',fontsize=23)

plt.axis('off')
plt.tight_layout(True)
plt.show()
```

<p style="text-align: center;"><span style='font-family: "Times New Roman", Times, serif; font-size: 21px;'>Above is an example of a graph with a maximum degree of 4 and a minimum of 2. <a id="1.4"></a>
<h3 style="background-color:orange;font-family:newtimeroman;font-size:200%;text-align:center;border-radius: 15px 50px;">Complement Of A Graph</h3>

<p style="text-align: center;"><span style='font-family: "Times New Roman", Times, serif; font-size: 24px;'>A compliment of graph $G(V, E)$ is defined as a graph with the same vertices. Still, all the edges connected in $G$ are no longer connected; instead, all the possible edges that were not connected in $G$ will be connected in the complement of $G$.</span></p>
<p style="text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><br></span></span></p>
<p style="text-align: center;"><span style='font-family: "Times New Roman", Times, serif; font-size: 24px;'>The formal definition; let $G(V, E)$ be an undirected graph the compliment of $G$ which is denoted by  $\overline{G}(V, E)$ is defined by the following rule: $(x,y) \in \overline{E} \Leftrightarrow x \neq y$ and $(x,y) \notin E$</span></p>

```python
fig = plt.figure(figsize=(14,14))

plt.subplot(1,2,1)
plt.title(r'Graph : $K_5$',fontsize=19,fontweight='bold')
plt.imshow(plt.imread('/kaggle/input/390-images-of-popular-graph-theory-graphs/Graph Images/g_K5.gif'))
plt.text(3,27,r'$V_1$',color='red',fontsize=23)
plt.text(33,6,r'$V_2$',color='red',fontsize=23)
plt.text(13,66,r'$V_3$',color='red',fontsize=23)
plt.text(53,66,r'$V_4$',color='red',fontsize=23)
plt.text(63,26,r'$V_5$',color='red',fontsize=23)


plt.axis('off')

plt.subplot(1,2,2)
plt.title(r'Graph : $\overline{K_5}$',fontsize=19,fontweight='bold')
plt.imshow(plt.imread('/kaggle/input/390-images-of-popular-graph-theory-graphs/Graph Images/g_5K1.gif'))
plt.text(3,27,r'$V_1$',color='red',fontsize=23)
plt.text(33,6,r'$V_2$',color='red',fontsize=23)
plt.text(13,66,r'$V_3$',color='red',fontsize=23)
plt.text(53,66,r'$V_4$',color='red',fontsize=23)
plt.text(63,26,r'$V_5$',color='red',fontsize=23)


plt.axis('off')
plt.show()
```

<p style="text-align: center;"><span style='font-family: "Times New Roman", Times, serif; font-size: 21px;'>Above is an example of a $K_5$ graph and its complement.</span></p>

<a id="2.1"></a>
<h3 style="background-color:orange;font-family:newtimeroman;font-size:200%;text-align:center;border-radius: 15px 50px;">The $K_n$ Family</h3>

<p style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; text-align: center;"><span data-preserver-spaces="true" style='color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; font-family: "Times New Roman", Times, serif; font-size: 24px;'>The $K$ graph family, also known as the &quot;whole graph&quot; family, defines all the graphs of size n (meaning they have n vertices) in which there is an edge between every pair of vertices.</span></p>
<p style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><br></span></span></p>
<p style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">Formal Definition;&nbsp;</span></span></span></p>
<ul style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">$K_n = (V,E)$</span></span></span></li>
    <li style="color: rgb(14, 16, 26); background: transparent; margin-top: 0pt; margin-bottom: 0pt; list-style-type: disc; text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;"><span data-preserver-spaces="true" style="color: rgb(14, 16, 26); background: transparent; margin-top:0pt; margin-bottom:0pt;">$V = \{v_i : 1\leq i\leq n\}$ $E =\{(v_i,v_j) = 1\leq i \leq j \leq n\}$</span></span></span></li>

<a id="3.1"></a>
<h3 style="background-color:orange;font-family:newtimeroman;font-size:200%;text-align:center;border-radius: 15px 50px;">Defenition Of A Clique</h3>

<p style="text-align: center;"><span style='font-family: "Times New Roman", Times, serif; font-size: 24px;'>Let $G(V, E)$ be an undirected graph.</span></p>
<p style="text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;">A set of vertices V&apos; in V &nbsp;is called a clique if every two vertices in the set are adjacent.</span></span></p>
<p style="text-align: center;"><span style='font-family: "Times New Roman", Times, serif; font-size: 24px;'>A clique set is called maximum if its cardinality is maximal among all independent sets in the graph ($|V'| \geq$ {set all of the cliques in $G$}) and is denoted by $\omega(G)$.</span></p>
<p><br></p>

```python
fig = plt.figure(figsize=(10,10))
plt.title(r'Graph : $\overline{X_{133}}$',fontsize=19,fontweight='bold')
plt.imshow(plt.imread('../input/390-images-of-popular-graph-theory-graphs/Graph Images/g_co-X133.gif'))
fig.text(.05,-.025,'$\omega(G)$ = 4',color='red',fontsize=23)
plt.text(3,46,r'$V_1$',color='red',fontsize=23)
plt.text(33,66,r'$V_2$',color='red',fontsize=23)
plt.text(3,86,r'$V_3$',color='red',fontsize=23)
plt.text(63,46,r'$V_4$',color='red',fontsize=23)
plt.text(63,86,r'$V_5$',color='red',fontsize=23)
plt.text(98,86,r'$V_6$',color='red',fontsize=23)
plt.text(33,6,r'$V_7$',color='red',fontsize=23)

plt.axis('off')

plt.show()
```

<h3 style="background-color:orange;font-family:newtimeroman;font-size:200%;text-align:center;border-radius: 15px 50px;">Defenition Of A Independent Set</h3>

<p style="text-align: center;"><span style='font-family: "Times New Roman", Times, serif; font-size: 24px;'>Let $G(V, E)$ be an undirected graph.</span></p>
<p style="text-align: center;"><span style="font-size: 24px;"><span style="font-family: 'Times New Roman', Times, serif;">A set of vertices $V' \in V$ &nbsp;is called independent if no two vertices in the set are adjacent.</span></span></p>
<p style="text-align: center;"><span style='font-family: "Times New Roman", Times, serif; font-size: 24px;'>An independent set is called maximum if its cardinality is maximal among all independent sets in the graph ($|V'| \geq ${all of the independent sets in G}) and is denoted by $\alpha(G)$.</span></p>
