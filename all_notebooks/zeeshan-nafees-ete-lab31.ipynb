```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
%matplotlib inline
import plotly.express as px
```

```python
df = pd.read_csv('/kaggle/input/indian-food-101/indian_food.csv')
df=df.replace(-1,np.nan)
df=df.replace('-1',np.nan)
df.head()
```

Q1):-Find percentage of vegetarian and non vegetarian. Show with suitable graph also.

```python
pie_df = df.diet.value_counts().reset_index()
pie_df.columns = ['diet','count']
fig = px.pie(pie_df, values='count', names='diet', title='Proportion of Vegetarian and Non-Vegetarian dishes',
             color_discrete_sequence=['green', 'red'])
fig.show()
```

Q5):-Compare preparation time and cooking time for Veg and Non Veg dishes. Represent using scatter plot or any other suitable plot.

```python
fig = px.scatter(df,x='cook_time',y='prep_time',color='diet',
                 color_discrete_sequence=['green','red'], hover_data = ['name'],
                 labels={
                     'cook_time': 'Cooking time (minutes)',
                     'prep_time': 'Preparation time (minutes)'
                 })
fig.show()
```
