# Which president had the largest age gap between them and their wife?

```python
lady_df = pd.read_csv("/kaggle/input/us-presidents-and-first-ladies/first_ladies.csv")
pres_df = pd.read_csv("/kaggle/input/us-presidents-and-first-ladies/presidents.csv")
```

```python
pres_df.head()
```

```python
lady_df.drop(lady_df.columns[0], axis=1, inplace=True)
lady_df.head()
```

```python
wife_df = lady_df.loc[lady_df['relation']=='Husband']
wife_df.head()
```

```python
# Find difference in ages
data = []
for idx, row in wife_df.iterrows():
    pres = row['president']
    wife_yr_born = row['born'].split(' ')[2]
    pres_row = pres_df.loc[pres_df['president'] == pres].iloc[0]
    pres_yr_born = pres_row['birth_date'].split(' ')[2]
    data.append((pres, pres_yr_born, row['name'], wife_yr_born))
    
```

```python
cols = ['president', 'pres_yr_born', 'first_lady', 'wife_yr_born']
df = pd.DataFrame(data, columns=cols)
def find_difference(row):
    return (int(row['wife_yr_born']) - int(row['pres_yr_born']))

differences = []
for idx, row in df.iterrows():
    differences.append(find_difference(row))
df['difference'] = differences
df.head()
```

```python
df.sort_values(by='difference', ascending=False)
```
