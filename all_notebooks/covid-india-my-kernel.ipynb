Data Cleaning

```python
df_covid_india.drop(["S. No."],axis=1,inplace=True)
```

```python
df_covid_india.rename(columns={"Name of State / UT":"State","Cured/Discharged/Migrated":"Discharged","Active Cases":"ActiveCases"},inplace=True)
```

```python
df_covid_india.head()
```

```python
    
df_covid_india.Deaths.plot(kind = 'line', color='g', label = 'Deaths', linewidth=1, alpha = 0.5, grid = True,linestyle = ':')
df_covid_india.Discharged.plot(color = 'r', label = 'Active Cases', linewidth=1, alpha = 0.5, grid = True, linestyle = '-.')
plt.legend(loc='upper right')
plt.xlabel('Deaths')
plt.ylabel('Active Cases')
plt.title('Line Plot')
```

**Total Discharged Patients**

```python

plt.figure(figsize=(18,10))
plt.bar(df_covid_india["State"],df_covid_india["Discharged"])
plt.title("Total Discharged Patients",fontsize=15)
plt.xlabel("Region",fontsize=14)
plt.ylabel("Patients",fontsize=14)
plt.xticks(rotation=90,size=14)
plt.yticks(size=14)
plt.show()
```

**Total Confirmed Cases**

```python
plt.figure(figsize=(18,10))
plt.bar(df_covid_india["State"],df_covid_india["Total Confirmed cases"])
plt.title("Total Deaths",fontsize=15)
plt.xlabel("Region",fontsize=14)
plt.ylabel("Patients",fontsize=14)
plt.xticks(rotation=90,size=14)
plt.yticks(size=14)
plt.show()
```

**Total Active Cases By state**

```python
plt.figure(figsize=(18,10))
plt.bar(df_covid_india["State"],df_covid_india["ActiveCases"])
plt.title("Total Active Cases By State",fontsize=15)
plt.xlabel("Region",fontsize=14)
plt.ylabel("Cases",fontsize=14)
plt.xticks(rotation=90,size=14)
plt.yticks(size=14)
plt.show()
```

**Death Rate**

```python
death_rate = df_covid_india['Deaths']/df_covid_india['Total Confirmed cases']*10
```

```python
plt.xlabel('State / Union terirorry', size=14,color='b')
plt.ylabel('Death Rate(%)', size=14,color='r')
plt.plot(df_covid_india['State'],death_rate)
plt.xticks(rotation='vertical' , color='indigo')
plt.yticks(np.arange(0,3.5,0.1))
plt.title('Death Rate', size = 20)
plt.grid(True)
```

**State wise Overview**

```python
df_covid_india1=df_covid_india.groupby("State")["ActiveCases","Deaths","Discharged","Total Confirmed cases"].sum()
df_covid_india1.head()
```

```python
m=df_covid_india1.loc["Bihar"]
plt.style.use("seaborn")
plt.title("Bihar Covid Overview")
labels=["Active Cases","Deaths","Discharged","Total Confirmed cases"]
colors=["#fc9303","#a22502","#00cd4e","#75ffed"]
plt.pie(m,labels=labels,colors=colors,wedgeprops={'edgecolor':'black'},autopct='%1.1f%%')
plt.show()
```

**India Covid19 Overview**

```python
Deaths=df_covid_india["Deaths"].sum()
Discharged=df_covid_india["Discharged"].sum()
Active=df_covid_india["ActiveCases"].sum()
Total_Cases=df_covid_india["Total Confirmed cases"].sum()
```

```python
slices=[Active,Deaths,Discharged,Total_Cases]
labels=["Active","Deaths","Discharged","Total_Cases"]
plt.title("India Covid19 OverView")
colors=["#fc9303","#a22502","#00cd4e","#75ffed"]
plt.pie(slices,colors=colors,autopct='%1.1f%%',labels=labels,shadow=True)
plt.show()
```
