```python
# import libraries and packages
import pandas as pd
import numpy as np
import requests
import sqlite3
import math
import PIL
import re
import urllib.request
import plotly.graph_objects as go
import IPython.display as Disp
from IPython.display import display_javascript, display_html, display, HTML, Image
from plotly.subplots import make_subplots
from glob import glob
from PIL import Image as IMG


# connect to the SQLite DB
conn = sqlite3.connect('../input/mubi_db.sqlite')


# define Python functions used to handle image data and display Plotly charts


def download_movie_image(movie_image_url, image_directory):
    urllib.request.urlretrieve(movie_image_url, image_directory)
    
    
def display_movie_image(image_data, movie_title, director_name, movie_release_year):
    
    display(Image(filename=image_data, retina=False))
    
    display(HTML('<strong>{}'.format(movie_title + ', ' + director_name + ' (' + str(movie_release_year) + ')')))

    
def display_plotly_line_charts(main_dataset, overall_title, first_subplot_title, second_subplot_title, third_subplot_title,
                               first_subplot_x_axis, first_subplot_y_axis, second_subplot_x_axis, second_subplot_y_axis,
                               third_subplot_x_axis, third_subplot_y_axis, first_subplot_y_axis_range, second_subplot_y_axis_range, 
                               third_subplot_y_axis_range, third_subplot_y_axis_scale, first_subplot_y_axis_tick_suffix, second_subplot_y_axis_tick_suffix, 
                               third_subplot_y_axis_tick_suffix, color_code, hover_template_first_subplot, hover_template_second_subplot,
                               hover_template_third_subplot):
    
    fig = make_subplots(
    rows=2, cols=2,
    specs=[[{}, {}],
           [{"colspan": 2}, None]],
    subplot_titles=(first_subplot_title, second_subplot_title, third_subplot_title),
    vertical_spacing=0.2)

    fig.add_trace(go.Scatter(x=main_dataset[first_subplot_x_axis], y=main_dataset[first_subplot_y_axis], name="",
                        hovertemplate = hover_template_first_subplot, line=dict(color=color_code)),
                 row=1, col=1)

    fig.add_trace(go.Scatter(x=main_dataset[second_subplot_x_axis], y=main_dataset[second_subplot_y_axis], name="",
                        hovertemplate = hover_template_second_subplot, line=dict(color=color_code)),
                 row=1, col=2)


    fig.add_trace(go.Scatter(x=main_dataset[third_subplot_x_axis], y=main_dataset[third_subplot_y_axis] * third_subplot_y_axis_scale, name="",
                        hovertemplate = hover_template_third_subplot, line=dict(color=color_code)),
                 row=2, col=1)

    fig.update_yaxes(range=first_subplot_y_axis_range, row=1, col=1, ticksuffix = first_subplot_y_axis_tick_suffix)

    fig.update_yaxes(range=second_subplot_y_axis_range, row=1, col=2, ticksuffix = second_subplot_y_axis_tick_suffix)

    fig.update_yaxes(range=third_subplot_y_axis_range, row=2, col=1, ticksuffix= third_subplot_y_axis_tick_suffix)


    fig.update_layout(showlegend=False, title_text=overall_title, title_x=0.5, font=dict(size=10))
    
    fig.show()
    
    
def display_plotly_bar_chart(main_dataset, hover_text_data, first_hover_text_value,
                                second_hover_text_value, third_hover_text_value,
                                fourth_hover_text_value, fifth_hover_text_value,
                                sixth_hover_text_value, seventh_hover_text_value,
                                eighth_hover_text_value, hover_text_value_types_list,
                                x_axis_column, y_axis_column, y_axis_range, marker_line_width,
                                bar_chart_title, x_axis_title, y_axis_title):
    
    hover_text = []
    
    for index, row in main_dataset.iterrows():
        
        hover_text.append(hover_text_data.format(hover_text_value_types_list[0](row[first_hover_text_value]),
                                                 hover_text_value_types_list[1](row[second_hover_text_value]),
                                                 hover_text_value_types_list[2](row[third_hover_text_value]),
                                                 hover_text_value_types_list[3](row[fourth_hover_text_value]),
                                                 hover_text_value_types_list[4](row[fifth_hover_text_value]),
                                                 hover_text_value_types_list[5](row[sixth_hover_text_value]),
                                                 hover_text_value_types_list[6](row[seventh_hover_text_value]),
                                                 hover_text_value_types_list[7](row[eighth_hover_text_value])))
        
        
    main_dataset['text'] = hover_text
    
    
    fig = go.Figure(data=[go.Bar(x = main_dataset[x_axis_column], y = main_dataset[y_axis_column],
            hovertext= hover_text)])

    fig.update_traces(marker_color = mubi_logo_color_code, marker_line_color = mubi_logo_color_code,
                  marker_line_width=marker_line_width)

    fig.update_layout(title_text= bar_chart_title,
                 title_x = 0.5, font=dict(size=14),
                 xaxis=dict(title = x_axis_title,
                            gridcolor='white',
                            type='log',
                            gridwidth=2),
                 yaxis=dict(title = y_axis_title,
                            gridcolor='white',
                            gridwidth=2,
                            range = y_axis_range
                            ))
    fig.show()
    

def display_plotly_bubble_chart(main_dataset, hover_text_data, first_hover_text_value,
                                second_hover_text_value, third_hover_text_value,
                                fourth_hover_text_value, fifth_hover_text_value,
                                sixth_hover_text_value, seventh_hover_text_value,
                                eighth_hover_text_value, bubble_dimension, bubble_size_column, 
                                bubble_color_dimension, bubble_color_column, x_axis_column, y_axis_column,
                                bubble_chart_title, x_axis_title, y_axis_title, y_axis_range):
    
    hover_text = []
    bubble_size = []
    
    for index, row in main_dataset.iterrows():
        
        hover_text.append(hover_text_data.format(row[first_hover_text_value],
                                                 row[second_hover_text_value],
                                                 row[third_hover_text_value],
                                                 row[fourth_hover_text_value],
                                                 row[fifth_hover_text_value],
                                                 row[sixth_hover_text_value],
                                                 row[seventh_hover_text_value],
                                                 row[eighth_hover_text_value]))
        
        bubble_size.append(math.sqrt(row[bubble_size_column]))
        
    main_dataset['text'] = hover_text
    main_dataset['size'] = bubble_size
    sizeref = 2.*max(main_dataset['size'])/(100**2)
    
    # Dictionary with dataframes for each bubble
    
    bubble_names = list(set(main_dataset[bubble_dimension]))
    bubble_data = {bubble:main_dataset.query(bubble_dimension + ' == ' + ' "{0}"'.format(bubble)) for bubble in bubble_names}
        
    # Create figure
    fig = go.Figure()

    for bubble_name, bubble in bubble_data.items():
        
        fig.add_trace(go.Scatter(
            x=bubble[x_axis_column], y=bubble[y_axis_column],
            name=bubble_data[bubble_name][bubble_color_dimension].values[0], text=bubble['text'],
            marker_size=bubble['size'], marker = dict(color = [bubble_data[bubble_name][bubble_color_column].values[0]]),
            showlegend=False
            ))
        
    # Tune marker appearance and layout
    fig.update_traces(mode='markers', marker=dict(sizemode='area',
                                              sizeref=sizeref, line_width=2))

    fig.update_layout(
    title = bubble_chart_title,
    xaxis=dict(
        title = x_axis_title,
        gridcolor='white',
        type='log',
        gridwidth=2
    ),
    yaxis=dict(
        title = y_axis_title,
        gridcolor='white',
        gridwidth=2,
        range = y_axis_range
    ),
    paper_bgcolor='rgb(243, 243, 243)',
    plot_bgcolor='rgb(243, 243, 243)',
    title_x = 0.5, font=dict(size=14))

    fig.show()
    

# define Mubi logo color code
mubi_logo_color_code = "#0508b8"


# extract movie data from 50 movies I love and that best represent the history of cinema 

history_of_cinema_fifty_film_ids_list = [164456, 373, 157887, 357, 18914, 3263, 214, 708, 723, 683, 
                                         300, 626, 255, 228, 450, 284, 295, 1934, 116671, 1435,
                                        823, 162, 18083, 187, 286, 510, 21884, 3494, 24187, 2105,
                                        26884, 103165, 200, 978, 20184, 26024, 1092, 29794, 2029, 1454,
                                        1103, 2465, 732, 2152, 12412, 455, 296, 270, 304, 418, 114074]


history_of_cinema_query = '''

WITH critics_table AS (

SELECT

    movie_id,
    
    COUNT(DISTINCT(rating_id)) number_of_critics    
    
FROM ratings

WHERE movie_id IN {0}

AND critic != "None"

GROUP BY 1

),


ratings_table AS (


SELECT

    movie_id,
    
    COUNT(DISTINCT(rating_id)) number_of_ratings,
    
    CAST(SUM(rating_score) AS FLOAT) / CAST(COUNT(DISTINCT(rating_id)) AS FLOAT) average_rating_score,
    
    CAST(SUM(rating_score) AS INT) total_rating_score

FROM ratings

WHERE movie_id IN {0}

GROUP BY 1

)


SELECT

    movies.movie_title,
    
    CAST(movies.movie_release_year AS INTEGER) movie_release_year,
    
    movies.director_name,
    
    movies.movie_popularity,
    
    movies.movie_image_url,
    
    critics_table.number_of_critics,
    
    ratings_table.number_of_ratings,
    
    ROUND(ratings_table.average_rating_score, 2) average_rating_score,
    
    ratings_table.total_rating_score
    
FROM movies

LEFT JOIN critics_table ON critics_table.movie_id = movies.movie_id

LEFT JOIN ratings_table ON ratings_table.movie_id = movies.movie_id

WHERE movies.movie_id IN {0}

ORDER BY 2 ASC


'''.format(tuple(history_of_cinema_fifty_film_ids_list))


history_of_cinema_data = pd.read_sql(history_of_cinema_query, conn)

for movie_number in range(0,len(list(history_of_cinema_data['movie_title']))):
    download_movie_image(history_of_cinema_data.sort_values(by='movie_release_year')['movie_image_url'][movie_number], "../movie_image_history" + str(movie_number) + ".jpg")
    
history_of_cinema_dictionary = {}
for movie in glob('../movie_image_history*'):
    history_of_cinema_dictionary[int(re.findall('\d+', movie)[0])] = IMG.open(movie)
```

## A) <a id='subpart_one_first_part'>Monthly Active Users & Monthly Paying Users</a>

```python
history_of_cinema_counter += 1

display_movie_image(history_of_cinema_dictionary[history_of_cinema_counter].filename, 
                    history_of_cinema_data['movie_title'][history_of_cinema_counter], 
                    history_of_cinema_data['director_name'][history_of_cinema_counter], 
                    history_of_cinema_data['movie_release_year'][history_of_cinema_counter])
```

```python
mau_query = '''


WITH DAU_TABLE AS (

SELECT

    DATE(rating_date_utc) date,

    COUNT(DISTINCT(user_id)) daily_active_users

FROM ratings_users

WHERE rating_date_utc >= '2020-01-01'

GROUP BY 1

),


MAU_TABLE AS (

SELECT 

    DAU_TABLE.date,

    DAU_TABLE.daily_active_users,

(SELECT

    COUNT(DISTINCT(user_id))

FROM ratings_users

WHERE JULIANDAY(DAU_TABLE.date) - 30 < JULIANDAY(DATE(rating_date_utc))

AND JULIANDAY(DATE(rating_date_utc)) <= JULIANDAY(DAU_TABLE.date)

) monthly_active_users

FROM DAU_TABLE

)


SELECT

    date,
    
    daily_active_users,
    
    monthly_active_users,
    
    CAST(daily_active_users AS FLOAT) / CAST(monthly_active_users AS FLOAT) stickiness
    
FROM MAU_TABLE

WHERE date < '2020-04-19'

ORDER BY 1

;

'''

mau_data = pd.read_sql(mau_query, conn)

display_plotly_line_charts(main_dataset = mau_data, overall_title = "Mubi Active Users Since January 2020", 
                           first_subplot_title = "Monthly Active Users <br> (30-Day Rolling Basis)", 
                           second_subplot_title = "Daily Active Users", third_subplot_title = "Stickiness",
                           first_subplot_x_axis = "date", first_subplot_y_axis = "monthly_active_users", second_subplot_x_axis = "date", second_subplot_y_axis = "daily_active_users",
                           third_subplot_x_axis = "date", third_subplot_y_axis = "stickiness", first_subplot_y_axis_range = [0, 53000], second_subplot_y_axis_range = [0, 7200], 
                           third_subplot_y_axis_range = [0, 20], third_subplot_y_axis_scale = 100, first_subplot_y_axis_tick_suffix = "", second_subplot_y_axis_tick_suffix = "", 
                           third_subplot_y_axis_tick_suffix = "%", color_code = mubi_logo_color_code, hover_template_first_subplot = 'MAU: %{y:,}<br> Date: %{x}', 
                           hover_template_second_subplot = 'DAU: %{y:,}<br> Date: %{x}', hover_template_third_subplot = "Stickiness: %{y:.2f}%<br> Date: %{x}")
```

Let's analyse here the retention rates for active users at Day 30.

In this part, a Mubi user is active if he rated one movie for a specific day or month.

The retention rate at Day 30 is defined by the following ratio: retained users at Day 30 (among the daily new users) / daily new active users


```python
retention_day_30_query = '''

WITH cohort_table AS 

(

SELECT

    user_id,

    MIN(rating_date_utc) cohort_date

FROM ratings_users 

GROUP BY 1

HAVING MIN(rating_date_utc) >= '2008-01-01'

ORDER BY 1

), 


retention_table AS 

(

SELECT

    cohort_date,
    
    rating_date_utc,
    
    cohort_table.user_id,
    
    (CASE WHEN JULIANDAY(rating_date_utc) - JULIANDAY(cohort_date) = 30 THEN 1 ELSE 0 END) is_retained_at_day_30

FROM ratings_users 

LEFT JOIN cohort_table ON ratings_users.user_id = cohort_table.user_id

WHERE cohort_date < rating_date_utc

GROUP BY 1,2,3

)

SELECT

    cohort_table.cohort_date,

    COUNT(DISTINCT(cohort_table.user_id)) daily_new_users,

    SUM(is_retained_at_day_30) retained_users_day_30,
    
    CAST(SUM(is_retained_at_day_30) AS FLOAT) / CAST(COUNT(DISTINCT(cohort_table.user_id)) AS FLOAT) retention_rate_day_30

FROM cohort_table

LEFT JOIN retention_table ON retention_table.user_id = cohort_table.user_id

GROUP BY 1

;

'''

retention_day_30_data = pd.read_sql(retention_day_30_query, conn)

retention_day_30_data = retention_day_30_data.fillna(0)

display_plotly_line_charts(main_dataset = retention_day_30_data, overall_title = "User Retention", 
                           first_subplot_title = "Daily New Users", 
                           second_subplot_title = "Retained Users at D+30", third_subplot_title = "Retention Rates at D+30",
                           first_subplot_x_axis = "cohort_date", first_subplot_y_axis = "daily_new_users", second_subplot_x_axis = "cohort_date", second_subplot_y_axis = "retained_users_day_30",
                           third_subplot_x_axis = "cohort_date", third_subplot_y_axis = "retention_rate_day_30", first_subplot_y_axis_range = [0, 2500], second_subplot_y_axis_range = [0, 30], 
                           third_subplot_y_axis_range = [0, 55], third_subplot_y_axis_scale = 100 ,first_subplot_y_axis_tick_suffix = "", second_subplot_y_axis_tick_suffix = "", 
                           third_subplot_y_axis_tick_suffix = "%", color_code = mubi_logo_color_code, hover_template_first_subplot = 'DNU: %{y:,}<br> Date: %{x}', 
                           hover_template_second_subplot = 'Retained Users D+30: %{y:,}<br> Date: %{x}', hover_template_third_subplot = "Retention Rate D+30: %{y:.2f}%<br> Date: %{x}")
```

## C) <a id='subpart_three_first_part'>Conversion Rates at D+60</a>

```python
history_of_cinema_counter += 1

display_movie_image(history_of_cinema_dictionary[history_of_cinema_counter].filename, 
                    history_of_cinema_data['movie_title'][history_of_cinema_counter], 
                    history_of_cinema_data['director_name'][history_of_cinema_counter], 
                    history_of_cinema_data['movie_release_year'][history_of_cinema_counter])
```

Since we do not have access to actual conversion data on Mubi, this KPI will be biased by the ratings data.

In this part, a Mubi user is a new trialist for a specific month if he rated his first movie on Mubi this month as a trialist. 

The user is converted at D+60 if he rated another movie within the next 60 days as a paying user.

On Mubi, new users are entitled to a 30-day free trial. Let's calculate the conversion rate at D+60 by cohort month.

Definition of the conversion rate: new Mubi trialists who are active paying users at D+60 (or before) / new Mubi trialists at start of period

```python
conversion_day_60_query = '''

WITH cohort_table AS 

(

SELECT

    user_id,

    MIN(rating_timestamp_utc) cohort_timestamp

FROM ratings

WHERE user_trialist = 1

GROUP BY 1

HAVING DATE(MIN(rating_timestamp_utc)) >= '2008-01-01'

ORDER BY 1

), 


conversion_first_table AS 

(

SELECT

    cohort_timestamp,
    
    rating_timestamp_utc,
    
    cohort_table.user_id,
    
    (CASE WHEN (JULIANDAY(DATE(rating_timestamp_utc)) - JULIANDAY(DATE(cohort_timestamp)) <= 60) THEN 1 ELSE 0 END) is_converted_at_day_60

FROM ratings

LEFT JOIN cohort_table ON ratings.user_id = cohort_table.user_id

WHERE cohort_timestamp < rating_timestamp_utc

AND user_trialist = 0

AND user_subscriber = 1

AND user_has_payment_method = 1

GROUP BY 1,2,3

HAVING (CASE WHEN (JULIANDAY(DATE(rating_timestamp_utc)) - JULIANDAY(DATE(cohort_timestamp)) <= 60) THEN 1 ELSE 0 END) = 1

),

conversion_second_table AS 

(

SELECT

    cohort_timestamp,
    
    user_id,
    
    MIN(rating_timestamp_utc) conversion_timestamp,
    
    is_converted_at_day_60

FROM conversion_first_table

GROUP BY 1,2

)


SELECT

    DATE(cohort_table.cohort_timestamp) cohort_date,

    COUNT(DISTINCT(cohort_table.user_id)) daily_new_trialists,

    SUM(is_converted_at_day_60) converted_users_day_60,
    
    CAST(SUM(is_converted_at_day_60) AS FLOAT) / CAST(COUNT(DISTINCT(cohort_table.user_id)) AS FLOAT) conversion_rate_day_60

FROM cohort_table

LEFT JOIN conversion_second_table ON conversion_second_table.user_id = cohort_table.user_id

GROUP BY 1

;

'''

conversion_day_60_data = pd.read_sql(conversion_day_60_query, conn)

conversion_day_60_data = conversion_day_60_data.fillna(0)
conversion_day_60_data['converted_users_day_60'] = conversion_day_60_data['converted_users_day_60'].astype(int)

display_plotly_line_charts(main_dataset = conversion_day_60_data, overall_title = "Conversions", 
                           first_subplot_title = "Daily New Trialists", 
                           second_subplot_title = "Converted Users at D+60", third_subplot_title = "Conversion Rates at D+60",
                           first_subplot_x_axis = "cohort_date", first_subplot_y_axis = "daily_new_trialists", second_subplot_x_axis = "cohort_date", second_subplot_y_axis = "converted_users_day_60",
                           third_subplot_x_axis = "cohort_date", third_subplot_y_axis = "conversion_rate_day_60", first_subplot_y_axis_range = [0, 750], second_subplot_y_axis_range = [0, 30], 
                           third_subplot_y_axis_range = [0, 120], third_subplot_y_axis_scale = 100, first_subplot_y_axis_tick_suffix = "", second_subplot_y_axis_tick_suffix = "", 
                           third_subplot_y_axis_tick_suffix = "%", color_code = mubi_logo_color_code, hover_template_first_subplot = 'New Trialists: %{y:,}<br> Cohort Date: %{x}', 
                           hover_template_second_subplot = 'Conversions D+60: %{y:,}<br> Cohort Date: %{x}', hover_template_third_subplot = "Conversion Rate D+60: %{y:.2f}%<br> Cohort Date: %{x}")
```

- The number of new trialists started increasing significantly on February 21<sup>st</sup> 2018 (40). 


- The number of daily new trialists and the number of converted users at D+60 both reached a peak during confinement in March - April 2020. 

  - On April 18<sup>th</sup>, Mubi attracted 712 new trialists.

```python
movie_popularity_by_release_date_query = '''

WITH release_year_table AS (

SELECT

    CAST(movie_release_year AS INT) movie_release_year,

    SUM(movie_popularity) overall_movie_popularity,

    COUNT(DISTINCT(movies.movie_id)) number_of_movies,

    CAST(CAST(SUM(movie_popularity) AS FLOAT) / CAST(COUNT(DISTINCT(movies.movie_id)) AS FLOAT) AS INT) average_movie_popularity
    
FROM movies

WHERE movie_release_year IS NOT NULL

GROUP BY 1

),

movie_ids_year_table AS (

SELECT

movie_id,

movie_release_year

FROM movies

WHERE movie_release_year IS NOT NULL

),


ratings_table AS (

SELECT

    movie_release_year,

    CAST(SUM(CASE WHEN rating_score IS NULL THEN 0 ELSE rating_score END) AS INT) total_score,
    
    COUNT(DISTINCT(rating_id)) number_of_ratings,
    
    ROUND(CAST(SUM(CASE WHEN rating_score IS NULL THEN  0 ELSE rating_score END) AS FLOAT) / CAST(COUNT(DISTINCT CASE WHEN rating_id IS NULL THEN 0 ELSE rating_id END) AS FLOAT),2) average_rating_score

FROM movie_ids_year_table a

LEFT JOIN ratings b ON b.movie_id = a.movie_id

GROUP BY 1

)


SELECT

    a.movie_release_year,

    overall_movie_popularity,

    number_of_movies,

    average_movie_popularity,

    total_score,
    
    number_of_ratings,
    
    average_rating_score

FROM release_year_table a

LEFT JOIN ratings_table b ON b.movie_release_year = a.movie_release_year

GROUP BY 1,2,3

ORDER BY 1

;

'''

movie_popularity_by_release_date_data = pd.read_sql(movie_popularity_by_release_date_query, conn)

movie_popularity_by_release_date_data['first_empty_column'] = ''
movie_popularity_by_release_date_data['second_empty_column'] = ''

hover_text_movie_popularity_by_release_date_data = ('Release Year: {0}<br>'+
                  'Average Movie Popularity: {1}<br>'+              
                  'Number of Movies: {2}<br>'+
                  'Overall Movie Popularity: {3}<br>'+
                  'Average Movie Rating Score: {4}<br>'+
                  'Number of Movie Ratings: {5}<br>')

display_plotly_bar_chart(movie_popularity_by_release_date_data, hover_text_data = hover_text_movie_popularity_by_release_date_data, 
                         first_hover_text_value = 'movie_release_year', second_hover_text_value = 'average_movie_popularity', third_hover_text_value = 'number_of_movies',
                         fourth_hover_text_value = 'overall_movie_popularity', fifth_hover_text_value = 'average_rating_score',
                         sixth_hover_text_value = 'number_of_ratings', seventh_hover_text_value = 'first_empty_column',
                         eighth_hover_text_value = 'second_empty_column', hover_text_value_types_list = [int, int, int, int, float, int, str, str],
                         x_axis_column = 'movie_release_year', y_axis_column = 'average_movie_popularity', y_axis_range = [0, 72], marker_line_width = 1.5,
                         bar_chart_title = 'Average Movie Popularity on Mubi by Release Date', x_axis_title = '', y_axis_title = '')
```

```python
top_50_directors_by_popularity_query = '''


WITH top_50_directors_by_movie_popularity_table AS (

SELECT

    director_name,
    
    director_id,
            
    SUM(movie_popularity) overall_movie_popularity,
    
    COUNT(DISTINCT(movie_id)) number_of_movies,
    
    CAST(SUM(movie_popularity) AS FLOAT) / CAST(COUNT(DISTINCT(movie_id)) AS FLOAT) average_movie_popularity
    
FROM movies

GROUP BY 1,2

ORDER BY 3 DESC

LIMIT 50

),


movie_ids_top_50_directors_by_movie_popularity_table AS (

SELECT

    a.movie_id,
    
    b.director_id,
    
    b.director_name,
    
    overall_movie_popularity,
    
    number_of_movies,
    
    average_movie_popularity

FROM movies a

INNER JOIN top_50_directors_by_movie_popularity_table b ON b.director_id = a.director_id

),


critics_top_50_directors_by_movie_popularity_table AS (

SELECT
    
    a.director_id,
    
    a.director_name,
        
    COUNT(DISTINCT(rating_id)) number_of_critics    
    
FROM movie_ids_top_50_directors_by_movie_popularity_table a

LEFT JOIN ratings b ON b.movie_id = a.movie_id

WHERE critic != "None"

GROUP BY 1,2

),


ratings_top_50_directors_by_movie_popularity_table AS (


SELECT
    
    a.director_id,
    
    a.director_name,
    
    COUNT(DISTINCT(rating_id)) number_of_ratings,
    
    SUM(rating_score) overall_rating_score,
    
    ROUND(CAST(SUM(b.rating_score) AS FLOAT) / CAST(COUNT(DISTINCT(b.rating_id)) AS FLOAT),2) average_rating_score
    
FROM movie_ids_top_50_directors_by_movie_popularity_table a

LEFT JOIN ratings b ON b.movie_id = a.movie_id

GROUP BY 1,2

)

SELECT

    a.director_name,
    
    a.overall_movie_popularity,
    
    a.number_of_movies,
        
    CAST(a.average_movie_popularity AS INT) average_movie_popularity,
        
    b.number_of_critics number_of_movie_critics,
    
    c.number_of_ratings number_of_movie_ratings,
    
    CAST(c.overall_rating_score AS INT) overall_movie_rating_score,
    
    c.average_rating_score average_movie_rating_score
    
FROM movie_ids_top_50_directors_by_movie_popularity_table a

LEFT JOIN critics_top_50_directors_by_movie_popularity_table b ON b.director_id = a.director_id

LEFT JOIN ratings_top_50_directors_by_movie_popularity_table c ON c.director_id = a.director_id

GROUP BY 1,2,3

ORDER BY 2 DESC

;

'''

top_50_directors_by_popularity_data = pd.read_sql(top_50_directors_by_popularity_query, conn)

top_50_directors_by_popularity_data['color'] = mubi_logo_color_code
top_50_directors_by_popularity_data['empty_column'] = ''
top_50_directors_by_popularity_data = top_50_directors_by_popularity_data.reset_index()
top_50_directors_by_popularity_data = top_50_directors_by_popularity_data.rename(columns={'index': 'rank'})
top_50_directors_by_popularity_data['rank'] += 1

hover_text_top_50_directors_by_popularity_data = ('Director: {0}<br>'+
                  'Rank: {1}<br>'+              
                  'Number of Movies: {2}<br>'+
                  'Overall Movie Popularity: {3}<br>'+
                  'Average Movie Popularity: {4}<br>'+
                  'Number of Movie Ratings: {5}<br>'+
                  'Average Movie Rating Score: {6}<br>'+
                  'Number of Movie Critics: {7}<br>')

display_plotly_bubble_chart(main_dataset = top_50_directors_by_popularity_data, hover_text_data = hover_text_top_50_directors_by_popularity_data, first_hover_text_value = 'director_name',
                                   second_hover_text_value = 'rank', third_hover_text_value = 'number_of_movies',
                                   fourth_hover_text_value = 'overall_movie_popularity', fifth_hover_text_value = 'average_movie_popularity',
                                   sixth_hover_text_value = 'number_of_movie_ratings', seventh_hover_text_value = 'average_movie_rating_score',
                                   eighth_hover_text_value = 'number_of_movie_critics',
                                   bubble_size_column = 'number_of_movie_ratings', bubble_dimension = 'director_name', bubble_color_dimension = 'empty_column',
                                   bubble_color_column = 'color', 
                                   x_axis_column = 'average_movie_rating_score', y_axis_column = 'overall_movie_popularity',
                                   bubble_chart_title = 'Top 50 Directors by Movie Popularity on Mubi', x_axis_title = 'Average Movie Rating Score', y_axis_title = 'Movie Popularity',
                                   y_axis_range = [0,79000])
```

```python
top_50_directors_by_movie_rating_score_query = '''


WITH directors_table AS (

SELECT

    director_name,
    
    director_id,
            
    SUM(movie_popularity) overall_movie_popularity,
    
    COUNT(DISTINCT(movie_id)) number_of_movies,
    
    CAST(SUM(movie_popularity) AS FLOAT) / CAST(COUNT(DISTINCT(movie_id)) AS FLOAT) average_movie_popularity
    
FROM movies

GROUP BY 1,2

ORDER BY 3 DESC

),


movie_ids_directors_table AS (

SELECT

    a.movie_id,
    
    b.director_id,
    
    b.director_name,
    
    overall_movie_popularity,
    
    number_of_movies,
    
    average_movie_popularity

FROM movies a

INNER JOIN directors_table b ON b.director_id = a.director_id

),


critics_directors_table AS (

SELECT
    
    a.director_id,
    
    a.director_name,
        
    COUNT(DISTINCT(rating_id)) number_of_critics    
    
FROM movie_ids_directors_table a

LEFT JOIN ratings b ON b.movie_id = a.movie_id

WHERE critic != "None"

GROUP BY 1,2

),


ratings_directors_table AS (


SELECT
    
    a.director_id,
    
    a.director_name,
    
    COUNT(DISTINCT(rating_id)) number_of_ratings,
    
    SUM(rating_score) overall_rating_score,
    
    ROUND(CAST(SUM(b.rating_score) AS FLOAT) / CAST(COUNT(DISTINCT(b.rating_id)) AS FLOAT),2) average_rating_score
    
FROM movie_ids_directors_table a

LEFT JOIN ratings b ON b.movie_id = a.movie_id

GROUP BY 1,2

)

SELECT

    a.director_name,
    
    a.overall_movie_popularity,
    
    a.number_of_movies,
        
    CAST(a.average_movie_popularity AS INT) average_movie_popularity,
        
    b.number_of_critics number_of_movie_critics,
    
    c.number_of_ratings number_of_movie_ratings,
    
    CAST(c.overall_rating_score AS INT) overall_movie_rating_score,
    
    c.average_rating_score average_movie_rating_score
    
FROM movie_ids_directors_table a

LEFT JOIN critics_directors_table b ON b.director_id = a.director_id

LEFT JOIN ratings_directors_table c ON c.director_id = a.director_id

WHERE c.number_of_ratings > 100

AND a.number_of_movies > 3

GROUP BY 1,2,3

ORDER BY 8 DESC, 7 DESC

LIMIT 50

;

'''

top_50_directors_by_movie_rating_score_data = pd.read_sql(top_50_directors_by_movie_rating_score_query, conn)

top_50_directors_by_movie_rating_score_data['color'] = mubi_logo_color_code
top_50_directors_by_movie_rating_score_data['empty_column'] = ''
top_50_directors_by_movie_rating_score_data = top_50_directors_by_movie_rating_score_data.reset_index()
top_50_directors_by_movie_rating_score_data = top_50_directors_by_movie_rating_score_data.rename(columns={'index': 'rank'})
top_50_directors_by_movie_rating_score_data['rank'] += 1

hover_text_top_50_directors_by_movie_rating_score_data = ('Director: {0}<br>'+
                  'Rank: {1}<br>'+              
                  'Number of Movies: {2}<br>'+
                  'Overall Movie Popularity: {3}<br>'+
                  'Average Movie Popularity: {4}<br>'+
                  'Number of Movie Ratings: {5}<br>'+
                  'Average Movie Rating Score: {6}<br>'+
                  'Number of Movie Critics: {7}<br>')

display_plotly_bubble_chart(main_dataset = top_50_directors_by_movie_rating_score_data, hover_text_data = hover_text_top_50_directors_by_movie_rating_score_data, first_hover_text_value = 'director_name',
                                   second_hover_text_value = 'rank', third_hover_text_value = 'number_of_movies',
                                   fourth_hover_text_value = 'overall_movie_popularity', fifth_hover_text_value = 'average_movie_popularity',
                                   sixth_hover_text_value = 'number_of_movie_ratings', seventh_hover_text_value = 'average_movie_rating_score',
                                   eighth_hover_text_value = 'number_of_movie_critics', bubble_size_column = 'number_of_movie_ratings', bubble_dimension = 'director_name', 
                                   bubble_color_dimension = 'empty_column', bubble_color_column = 'color', x_axis_column = 'average_movie_rating_score', 
                                   y_axis_column = 'overall_movie_popularity', bubble_chart_title = 'Top 50 Directors by Movie Rating Score on Mubi', 
                                   x_axis_title = 'Average Movie Rating Score', y_axis_title = 'Movie Popularity', y_axis_range = [-8000,83000])
```
