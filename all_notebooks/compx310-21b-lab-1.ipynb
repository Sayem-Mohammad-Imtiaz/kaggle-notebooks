```python
X= dataframe.iloc[:,1:-1]
X
Y= dataframe.iloc[:,-1]
Y
```

X= dataframe.iloc[:,1:-1] is doing the second coloum to the last coloum

Y= dataframe.iloc[:,-1] is counting the last coloum


```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y, test_size =0.2, random_state=1506360)
```

Split the dataset into 20% (test): 80%(train) ratio

```python
from sklearn.svm import SVC
A=SVC()
A
A.fit(X_train, Y_train)
```

SVC is a default hyperparameters

```python
P=A.predict(X_test)
P
```

A.predict is listing all 0 and 1 value of the X_test

```python
Y_test[:5]
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(P,Y_test)
```

It prints out the confusion matrix for P and Y_test

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, P))
```
