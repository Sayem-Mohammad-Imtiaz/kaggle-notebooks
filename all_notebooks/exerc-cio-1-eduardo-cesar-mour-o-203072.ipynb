### Questão 1 - Item C - Representação Gráfica
Para cada uma das variáveis, produza um ou mais gráficos, usando matplotlib, que descreva seu comportamento / caracteristica. Lembre-se que estes gráficos precisam ser compatíveis com a classificação da variável. 

```python
#Gráfico 1 - Agrupando df para total de eleitores por UF
x = df[['uf','total_eleitores']].groupby('uf',as_index = False).sum().sort_values(by = 'total_eleitores', ascending = False)
x
```

```python
#Plotando Plotly gráfico 1
fig = go.Figure(
    data=[go.Bar(x=x['uf'], y=x['total_eleitores'])],
    layout=go.Layout(
        title=go.layout.Title(text="Eleitores por estado")
    )
)

py.iplot(fig)
```

```python
#Gráfico 2 - % de eleitores por gênero
masculino = df['gen_masculino'].sum()
feminino = df['gen_feminino'].sum()
n_if = df['gen_nao_informado'].sum()
```

```python
#Plotando Plotly gráfico 2
langs = ['Masculino', 'Feminino', 'Não Informado']
Valores = [masculino, feminino, n_if]
trace = go.Pie(labels = langs, values = Valores)
data = [trace]
fig = go.Figure(data = data, layout=go.Layout(title=go.layout.Title(text="% de Eleitores Totais por Gênero")))
py.iplot(fig)
```

```python
#Gráfico 3 - Agrupando faixa etária em uma series e depois somando o total por categoria
idade =  np.array(['f_16', 'f_18_20', 'f_25_34', 'f_45_59', 'f_70_79', 'f_sup_79'])

tot_idade = np.array([df[i].sum() for i in idade])

tot_idade
```

```python
#Plotando Plotly gráfico 3
fig = go.Figure(
    data=[go.Bar(x = idade, y = tot_idade)],
    layout=go.Layout(
        title=go.layout.Title(text="Quantidade de Eleitores p/ Faixa Etária")
    )
)

py.iplot(fig)
```

```python
#Gráfico 4 - Agrupando total de eleitores por município e pegando os top 5 em quantidade de eleitores
a = df[['nome_municipio','total_eleitores']].groupby('nome_municipio',as_index = False).sum().sort_values(by = 'total_eleitores', ascending = False).head(5)
a
```

```python
#Criando distribuição de eleitores por gênero para o estado de SP
df1 = df[df['nome_municipio']=='SÃO PAULO']
eleitor_tot = df1['total_eleitores']
masculinosp = df1['gen_masculino']
femininosp = df1['gen_feminino']
nifsp = df1['gen_nao_informado']
```

```python
#Teste
eleitor_tot
```

```python
#Plotando gráfico 4
generos =['Masculino', 'Feminino', 'Não Informado']

fig = go.Figure(data=[
    go.Bar(name='Masculino', x=eleitor_tot, y = masculinosp),
    go.Bar(name='Feminino', x=eleitor_tot, y = femininosp),
    go.Bar(name='Não Informado', x=eleitor_tot, y = nifsp),
],layout=go.Layout(title=go.layout.Title(text="Distribuição de Quantidade de Eleitores p/ Gênero no Município de SP")))
# Change the bar mode
fig.layout.update(barmode='stack')
py.iplot(fig)
```

```python
#Gráfico 5 - Agrupando quantidade de municípios por estado
z = df[['uf','nome_municipio']].groupby('uf', as_index = False).count().sort_values(by = 'nome_municipio', ascending = False)
z
```

```python
#Plotando gráfico 5
fig = go.Figure(
    data=[go.Bar(x=z['uf'], y=z['nome_municipio'])],
    layout=go.Layout(
        title=go.layout.Title(text="Quantidade de município por estado")
    )
)

py.iplot(fig)
```

```python
#Gráfico 6 - Agrupando total de eleitores por estado
eleit_est = df[['uf','total_eleitores']].groupby('uf',as_index = False).sum().sort_values(by = 'total_eleitores', ascending = False)
eleit_est
```

```python
#Criando variável de total de eleitores
total = eleit_est['total_eleitores'].sum()
```

```python
#Atribuindo divisão de eleitores por estado pelo total para encontrar %
percentage = eleit_est['total_eleitores'].div(total).mul(100)
percentage.count()
```

```python
#Teste
percentage
```

```python
#Plotando gráfico 6
langs1 = ['SP', 'MG', 'RJ', 'BA', 'RS', 'PR', 'PE', 'CE', 'PA', 'SC', 'MA', 'GO', 'PB', 'ES', 'RN', 'PI', 'AM', 'MT', 'AL', 'MS', 'SE', 'RO' , 'TO', 'AC', 'AP', 'RR']
Valores1 = [values for values in percentage]
trace1 = go.Pie(labels = langs1, values = Valores1)
data1 = [trace1]
fig1 = go.Figure(data = data1, layout=go.Layout(title=go.layout.Title(text="% do total de eleitores por estado")))
py.iplot(fig1)
```
