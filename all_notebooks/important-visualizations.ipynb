# All visualizations are Mouse Hover figures. Don't forget to bring your cursor over the visualizations

```python
df=pd.read_csv('../input/palestine-body-count/Palestine Body Count.csv')
```

```python
pd.isnull(df).sum()
```

> Number Of instances by Year

```python

DF= pd.DataFrame(df.groupby("Year")["Palestinians Killed"].max().sort_values(ascending = False))
DF.reset_index(level=0,inplace=True)
fig = px.pie(DF,
             values='Palestinians Killed',
             names='Year',
             hover_data=['Palestinians Killed'],
             labels={'country':'country'}
            )
fig.update_layout(title='<b>Palestinians Killed By Year Hover Pie-chart<b>',
                  titlefont={'size': 24,
                             'family': 'Serif',
                            },
                  showlegend=False, 
                  width=600, 
                  height=600)
fig.show()
```

# >Bring your cursor over pie-chart  

```python
df1= pd.DataFrame(df.groupby("Year")["Palestinians Killed"].max().sort_values(ascending = False))
df2= pd.DataFrame(df.groupby("Year")["Israelis Injuries"].max().sort_values(ascending = False))
df3= pd.DataFrame(df.groupby("Year")["Israelis Killed"].max().sort_values(ascending = False))
df4=pd.concat([df1,df2, df3],  axis=1)
df4.reset_index(level=0, inplace=True)
```

```python
import plotly.graph_objs as go
from plotly.offline import init_notebook_mode, iplot, plot

trace1 = go.Scatter(x = df4['Year'],
                    y = df4['Palestinians Killed'],
                    mode = "lines+markers",
                    name = "Palestinians Killed",
                    marker = dict(color = 'red'),
                    text= df4['Year'])
trace2 = go.Scatter(x = df4['Year'],
                    y = df4['Israelis Injuries'],
                    mode = "lines+markers",
                    name = "Israelis Injuries",
                    marker = dict(color = 'blue'),
                    text= df4['Year'])
trace3 = go.Scatter(x = df4['Year'],
                    y = df4['Israelis Killed'],
                    mode = "lines+markers",
                    name = "Israelis Killed",
                    marker = dict(color = 'green'),
                    text= df4['Year'])

data = [trace1, trace2, trace3]
layout = dict(title = ' Comparision of casualties throughout ',
              xaxis= dict(title= 'Year',ticklen= 5,zeroline= False),
               
              hovermode="x unified")
fig = dict(data = data, layout = layout)
iplot(fig)     
```
