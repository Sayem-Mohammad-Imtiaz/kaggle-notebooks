```python
df_life = pd.read_csv('/kaggle/input/life-expectancy-global-trend/life_expectancy_years.csv')
df_country = pd.read_csv('/kaggle/input/life-expectancy-global-trend/countries_total.csv')
df_population = pd.read_csv('/kaggle/input/life-expectancy-global-trend/population_total.csv')
```

```python
# Which country has the largest life expectancy in 2018?
def largest_life():
    result = df_life.loc[df_life['2018'].idxmax()]['geo']
    return result
largest_life()
```

```python
# Which country has the smallest life expectancy in 2018?
def min_life():
    result = df_life.loc[df_life['2018'].idxmin()]['geo']
    return result
min_life()
```

```python
# In the first decile, how old is the highest life expectancy by threshold ?
def life_first_threshold():
    df = df_life['2018'].dropna()
    var = np.percentile(df,90)
    result = round(var,1)
    return result
life_first_threshold()
```
