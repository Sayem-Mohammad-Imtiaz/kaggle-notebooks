```python
data=data.drop('peakrpm',axis=1)

data=data.drop('stroke',axis=1)

data=data.drop('wheelbase',axis=1)

data=data.drop('boreratio',axis=1)

data=data.drop('compressionratio',axis=1)

data=data.drop('drivewheel',axis=1)

data=data.drop('doornumber',axis=1)

data=data.drop('fuelsystem',axis=1)

data=data.drop('enginetype',axis=1)
```

```python
data.head()
```

```python
sig_col = ['CarName','fueltype', 'aspiration','carbody','cylindernumber','curbweight', 'enginesize','horsepower', 
                    'citympg','highwaympg', 'carlength','carwidth','symboling','enginelocation']

X=data[sig_col]
Y=data['price']
X.head()
```

```python
#For Carbody

def convert_to_int(word):
    word_dict = {'convertible':1, 'hatchback':2, 'sedan':3, 'wagon':4, 'hardtop':5, 0: 0}
    return word_dict[word]

X['carbody'] = X['carbody'].apply(lambda x : convert_to_int(x))


#For cylinder number
def convert_to_int(word):
    word_dict = {'two':2, 'three':3, 'four':4, 'five':5, 'six':6,'eight':8,'twelve':12,0: 0}
    return word_dict[word]

X['cylindernumber'] = X['cylindernumber'].apply(lambda x : convert_to_int(x))



#For Carname
def convert_to_int(word):
    word_dict = {'alfa-romero':1, 'audi':2, 'bmw':3, 'chevrolet':4, 'dodge':5,'honda':6,'isuzu':7,'jaguar': 8,'mazda':9,
                'buick':10,'mercury':11,'mitsubishi':12,'Nissan':13,'peugeot':14,'plymouth':15,'porsche':16,'renault':17,
                'saab':18,'subaru':19,'toyota':20,'volkswagen':21,'volvo':22}
    return word_dict[word]

X['CarName'] = X['CarName'].apply(lambda x : convert_to_int(x))



#FOR FuelType
def convert_to_int(word):
    word_dict = {'gas':1, 'diesel':2}
    return word_dict[word]

X['fueltype'] = X['fueltype'].apply(lambda x : convert_to_int(x))


#FOR aspiration
def convert_to_int(word):
    word_dict = {'std':1, 'turbo':2}
    return word_dict[word]

X['aspiration'] = X['aspiration'].apply(lambda x : convert_to_int(x))


#For engine location
def convert_to_int(word):
    word_dict = {'front':1, 'rear':2}
    return word_dict[word]

X['enginelocation'] = X['enginelocation'].apply(lambda x : convert_to_int(x))


X.head()
```

```python
Y.head()
```

```python
import sklearn
import pickle
from sklearn import linear_model
```

```python
# TRAIN MODEL MULTIPLE TIMES FOR BEST SCORE
best = 0
for _ in range(20):
    x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(X, Y, test_size=0.1)

    linear = linear_model.LinearRegression()

    linear.fit(x_train, y_train)
    acc = linear.score(x_test, y_test)
   # print("Accuracy: " + str(acc))

    if acc > best:
        best = acc
        with open("carprices.pickle", "wb") as f:
            pickle.dump(linear, f)


pickle_in = open("carprices.pickle", "rb")
linear = pickle.load(pickle_in)
acc
```

```python
predicted= linear.predict(x_test)
x_test1=np.array(x_test)
y_test1=np.array(y_test)
```

```python
#Comparing predicitons with actual results
for w in range(21):
    print(y_test1[w],' ',x_test1[w])
    print(predicted[w])
    print()
```
