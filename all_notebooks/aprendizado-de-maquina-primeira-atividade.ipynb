```python
data = pd.read_csv("/kaggle/input/predict-diabetes-based-on-diagnostic-measures/diabetes.csv")
print("Características dos atributos")
print(data.info())
print("\nMédia dos atributos")
print(data.mean())
print("\nMediana dos atributos")
print(data.median())
```

```python
sample_30_no_replacement = data.sample(frac = 0.3)
print("Atributos de amostragem simples de 30% sem reposição")
print("Características dos atributos")
print(sample_30_no_replacement.info())
print("\nMédia dos atributos")
print(sample_30_no_replacement.mean())
print("\nMediana dos atributos")
print(sample_30_no_replacement.median())
```

```python
sample_30_with_replacement = data.sample(frac = 0.3, replace = True)
print("Atributos de amostragem simples de 30% com reposição")
print("Características dos atributos")
print(sample_30_with_replacement.info())
print("\nMédia dos atributos")
print(sample_30_with_replacement.mean())
print("\nMediana dos atributos")
print(sample_30_with_replacement.median())
```

```python
sample_50_no_replacement = data.sample(frac = 0.5)
print("Atributos de amostragem simples de 50% sem reposição")
print("Características dos atributos")
print(sample_50_no_replacement.info())
print("\nMédia dos atributos")
print(sample_50_no_replacement.mean())
print("\nMediana dos atributos")
print(sample_50_no_replacement.median())
```

```python
sample_50_with_replacement = data.sample(frac = 0.5, replace = True)
print("Atributos de amostragem simples de 50% com reposição")
print("Características dos atributos")
print(sample_50_with_replacement.info())
print("\nMédia dos atributos")
print(sample_50_with_replacement.mean())
print("\nMediana dos atributos")
print(sample_50_with_replacement.median())
```

```python
classes = data['diabetes'].unique()

labelFrequencies = {}
for clz in classes:
    labelFrequencies[clz] = int(data[data.diabetes == clz].shape[0])
print(labelFrequencies)
```

```python
frac = 0.5
newLabelFrequencies = {label: int(frequency * frac) for (label,frequency) in labelFrequencies.items()}
print(newLabelFrequencies)
samplesPerClass = []
for c in classes:
    currentClassData = data[data['diabetes'] == c]
    currentClassSamples = currentClassData.sample(n = newLabelFrequencies[c])
    samplesPerClass.append(currentClassSamples)
stratifiedSamples50 = pd.concat(samplesPerClass)

print("Atributos de amostragem estratificada de 50%")
print("Características dos atributos\n")
print(stratifiedSamples50.info())
print("\nMédia dos atributos")
print(stratifiedSamples50.mean())
print("\nMediana dos atributos")
print(stratifiedSamples50.median())
```
