```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
test_size=int(len(data1)*0.2)
print("total records =",len(data1),"test size =",test_size)
train_set=data1[:-test_size]
test_set=data1[-test_size:]
print(train_set.shape)
print(test_set.shape)
```

```python
train_set_x=train_set["Fuel Consumption Comb (L/100 km)"]
train_set_x=train_set_x[:,np.newaxis]
print(train_set_x)
```

```python
train_set_y=train_set["CO2 Emissions(g/km)"]
train_set_y=train_set_y[:,np.newaxis]
print(train_set_y)
```

```python
test_set_x=test_set["Fuel Consumption Comb (L/100 km)"]
test_set_x=test_set_x[:,np.newaxis]
test_set_y=test_set["CO2 Emissions(g/km)"]
test_set_y=test_set_y[:,np.newaxis]



print(train_set_x.shape)
print(train_set_y.shape)
print(test_set_x.shape)
print(test_set_y.shape)
```

```python
lm = LinearRegression()
lm.fit(train_set_x,train_set_y)
```

```python
y_train_pred=lm.predict(train_set_x)
y_test_pred=lm.predict(test_set_x)
```

```python
plt.scatter(train_set_x,train_set_y)
plt.scatter(train_set_x,y_train_pred)
plt.scatter(test_set_x,y_test_pred,marker=".")
```

```python
newval=pd.DataFrame({'cons':[12]})
p1=lm.predict(newval)
print(p1)
```

```python
y_train_pred=lm.predict(train_set_x)
mean_absolute_error(train_set_y,y_train_pred),mean_squared_error(train_set_y,y_train_pred)
```

```python
y_test_pred=lm.predict(test_set_x)
mean_absolute_error(test_set_y,y_test_pred),mean_squared_error(test_set_y,y_test_pred)
```
