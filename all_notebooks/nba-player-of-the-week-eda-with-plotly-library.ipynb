Let's change Date format to datetime to use easily.

```python
data.Date = pd.to_datetime(data.Date)
```

```python
data.Position = data.Position.str.replace('-','')
# Histogram of position
hist_data = [go.Histogram(x =data.Position)]
layout = go.Layout(
    title='Histogram of Positon',
    xaxis=dict(
        title='Position'
    ),
    yaxis=dict(
        title='Count'
    ),
)
fig = go.Figure(data=hist_data, layout=layout)
iplot(fig)
```

Guards are the players who selected player of the week most.

```python
#scatter graph of age and season in league
trace = go.Scatter(x = data.Age,
                   y = data['Seasons in league'],
                   mode = 'markers',
                   text = data.Team)
layout = go.Layout(
    title='Age vs Season in league',
    xaxis=dict(
        title='Age'
    ),
    yaxis=dict(
        title='Seasons in league'
    ),
)
scatter_data = [trace]
fig = go.Figure(data=scatter_data, layout=layout)
iplot(fig)
```

As age increases seasons in the league is also increasing as expected. There is an interesting data point in the age 31 and first season in the league.
Age range of the player of the week is also easily seen on the graph which distributed between 19 and 40.

```python
#pie chart of the conferences.
fig = {
  "data": [
    {
      "values": data.Conference.value_counts().tolist(),
      "labels": data.Conference.value_counts().keys().tolist(),
      "domain": {"x": [0, .5]},
      "hoverinfo":"label+percent+value",
      "hole": .3,
      "type": "pie"
    },],
  "layout": {
        "title":"Percentage of Conferences"
            },
    }
iplot(fig)
```

Quite even distribution by the conferences.

```python
# bubble plot of age of player of the week by time. Size of the bubble is weight of the player
# color of the bubble is season
bubble_data = [
    {
        'y': data.Age,
        'x': data.Date,
        'mode': 'markers',
        'marker': {
            'color': data['Season short'],
            'size': data.Weight/10,
            'showscale': True
        },
        "text" :  data.Player    
    }
]
iplot(bubble_data)
```

```python
#box plot of weight by position

x_data = data.Position.unique()
y_data = []
text_data = []
for each in x_data:
    y_data.append(data[data.Position == each].Weight)
    text_data.append(data[data.Position == each].Player)
    
x_data = np.append(x_data,'total')

y_data.append(data.Weight)
text_data.append(data.Player)
#y_data = y_data.tolist()

traces = []

for xd, yd, td in zip(x_data, y_data, text_data):
        traces.append(go.Box(
            y=yd,
            name=xd,
            boxpoints='all',
            jitter=0.5,
            text = td,
            whiskerwidth=0.2,
            marker=dict(
                size=2,
            ),
            line=dict(width=1),
        ))

layout = go.Layout(
    title='Weight of player of the week by position',
    yaxis=dict(
        autorange=True,
        showgrid=True,
        zeroline=True,
        dtick=5,
        gridcolor='rgb(255, 255, 255)',
        gridwidth=1,
        zerolinecolor='rgb(255, 255, 255)',
        zerolinewidth=2,
    ),
    margin=dict(
        l=40,
        r=30,
        b=80,
        t=100,
    ),
    paper_bgcolor='rgb(243, 243, 243)',
    plot_bgcolor='rgb(243, 243, 243)',
    showlegend=False
)

fig = go.Figure(data=traces, layout=layout)
iplot(fig)
```

Centers are biggest players and guards are the 'tiniest'.
