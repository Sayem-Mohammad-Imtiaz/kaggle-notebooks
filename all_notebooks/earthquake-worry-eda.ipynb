## Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
data_set = pd.read_csv("/kaggle/input/earthquake-worry/earthquake_data.csv")
```

```python
data_set.head(5)
```

```python
data_set.describe(include = 'all')
```

```python
data_set.info()
```

```python
data_set.isnull().sum()
```

```python
data_set = data_set.rename(columns = {'In general, how worried are you about earthquakes?':'worried', 'How worried are you about the Big One, a massive, catastrophic earthquake?': 'how worried?',
                                      'catastrophic earthquake?': 'catastrophic', 'Do you think the "Big One" will occur in your lifetime?': 'lifetime',
                                      'Have you ever experienced an earthquake?': 'experienced', 'Have you or anyone in your household taken any precautions for an earthquake (packed an earthquake survival kit, prepared an evacuation plan, etc.)?': 'precaution',
                                      'How familiar are you with the San Andreas Fault line?': 'San Andreas fault', 'How familiar are you with the Yellowstone Supervolcano?': 'Yellowstone Supervolcano',
                                      'What is your gender?': 'gender', 'How much total combined money did all members of your HOUSEHOLD earn last year?': 'income last year', 'Age': 'age'})
```

```python
data_set.head(5)
```

## Exploratory Data Analysis

```python
plt.figure(figsize = (12, 8))
sns.countplot(x = data_set['worried'], data = data_set)
plt.title("Reactions towards Earthquake and Volcanoes")
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.countplot(x = data_set['lifetime'], data = data_set)
plt.title("Reactions to the question: Do you think the 'Big One' will occur in your lifetime?")
plt.show()
```

```python
plt.figure(figsize=(30,10))
plt.subplots_adjust(left=0.125, bottom=0.1, right=0.9, top=0.9,
                      wspace=0.5, hspace=0.2)
plt.subplot(141)
plt.title('US Regions')
data_set['US Region'].value_counts().plot.pie(autopct="%1.1f%%")
```

```python
plt.figure(figsize = (12, 8))
sns.countplot(x = data_set['age'], data = data_set)
plt.title("Age")
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.countplot(y = data_set['San Andreas fault'], data = data_set)
plt.title("Idea about San Andreas Fault")
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.countplot(y = data_set['Yellowstone Supervolcano'], data = data_set)
plt.title("Idea about Yellowstone Supervolcano")
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.countplot(y = data_set['income last year'], data = data_set)
plt.title("Annual Income (Previous Year)")
plt.show()
```

```python
fig = px.pie(data_frame = data_set, names='income last year', title = 'Income', hole = 0.2)
fig.show()
```

```python
fig = px.pie(data_frame = data_set, names='age', title = 'Income', hole = 0.2)
fig.show()
```
