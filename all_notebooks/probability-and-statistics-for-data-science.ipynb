3. Conditional Probability - a measure of calculating a probablity of an event (A) to occur given that another event (B) has already occured.

The formula for conditional probability is P(A|B) = P(A ∩ B) / P(B)

P(A|B) = probability of A occurring, given B occurs 

P(A ∩ B) = probability of both A and B occurring

P(B) = probability of B occurring

```python
# Conditional probability with python example

len(loan_train_data)   #checking the number of records
```

```python
# We’re going to calculate the probability an applicant gets an approval of loan, 
# given that their credit history is good.

pd.crosstab(loan_train_data['Credit_History'], loan_train_data['Loan_Status'], margins = True)
```

*  P(A): Loan has been approved
*  P(B): Credit history is good (1)
*  P(A|B): Getting loan approval provided that credit history is good

**According to the formula,**

* P(A) = 7 + 415 / (110 + 415 + 82 + 7) = 0.687296417
* P(B) = 110 + 415 / (110 + 415 + 82 + 7) = 0.85504886
* P(A ∩ B) = 415/(110 + 415 + 82 + 7) = 0.675895765
* P(A|B) = 0.675895765/0.85504886 = 0.79047619

**The probability of getting a loan approval, given that the credit history is good  is 79%.**









**4. Mean, Median, Mode, Standard Deviation, Variance**

* Mean - mean of a dataset is calculated by adding all the numbers in the dataset and dividing it by the total number of values in that dataset.

* Median - Median is the value at the centre of dataset after arranging it in ascending order.

* Mode - Mode is the value that occurs most often in a dataset.

* Standard Deviation - a quantity that tells amount of variation/dispersion within a dataset. It gives an idea about how much the items in the dataset differ from the mean of the datset. 

* Variance - a qauntity that describes how far the numbers are spread out from their mean.

```python
print(loan_train_data.describe())
print(loan_train_data['LoanAmount'].describe())


mean = loan_train_data['LoanAmount'].mean()
median = loan_train_data['LoanAmount'].median()
mode = loan_train_data['LoanAmount'].mode()[0]
std = loan_train_data['LoanAmount'].std()
var = loan_train_data['LoanAmount'].var()

print('\n')
print('Mean =',mean)
print('Median =',median)
print('Mode =',mode)
print('Standard Deviation =',std)
print('Variance =',var)
```

```python
# Implementing different distributions in Python

# 1. Normal Distribution in Python

from numpy import random

normaldis = random.normal(loc = 3, scale = 2, size = (1000))


sns.distplot(normaldis, hist = True)
plt.show()
```

```python
# 2. Bionomial Distribution

sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=True)

plt.show()
```

```python
# 3. Poisson Distribution

sns.distplot(random.poisson(lam=2, size=1000), kde=True)

plt.show()
```

```python
# 4. Uniform Distribution

sns.distplot(random.uniform(0.1, 10, size=1000), hist=True)

plt.show()
```

```python
# 5. Logistic Distribution

sns.distplot(random.logistic(2, 1, size=1000), hist=False)

plt.show()
```
