```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
import time
```

```python
data = pd.read_csv("../input/breast-cancer-wisconsin-data/data.csv",header=0) #read data from kaggle
```

```python
data.head() #first 5 row
```

```python
data.drop(["id","Unnamed: 32"],inplace=True,axis=1) #unnecessary columns drop
```

```python
data.isnull().sum()  # there are not any isnull
```

```python
data.dtypes #all variable float64 , not including target column(diagnosis)
```

```python
data.describe().T #describe statistic values
```

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
la=LabelEncoder()
data["target"]=la.fit_transform(data.diagnosis)
#in here, we add new column(target) and this column consist 0 or 1 ; B=0 , M=1
```

```python
#We shoul drop old target(diagnosis)
data.drop(["diagnosis"],inplace=True,axis=1) 
```

```python
data.head()
```

```python
X =data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X)
sc.transform(X_train)
sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
```

```python
y_preds=knn.predict(X_test)
```

```python
sample=pd.DataFrame({"Y_predicted":y_preds,
                    "Real_Y_test":y_test})
sample
```

```python
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
```

```python
acc = accuracy_score(y_test,y_preds)
print(acc) #basic accuracy score is %95.9
```

```python
#now we should improve us score
#Model tuning;
knn = KNeighborsClassifier()
knn_params = {"n_neighbors": np.arange(1,50),
              "weights":["uniform","distance"]}

knn_cv = GridSearchCV(knn, knn_params, cv=5).fit(X_train, y_train)
print("Best Parameters: " + str(knn_cv.best_params_))
```

```python
knn_tuned = KNeighborsClassifier(n_neighbors=11,weights="distance").fit(X_train,y_train)
```

```python
new_pred=knn_tuned.predict(X_test)
new_acc=accuracy_score(y_test,new_pred)
print(new_acc)
```
