**Loading dataset**

```python
df = pd.read_csv('../input/shopping-annual-income/shopping_data.csv')
```

**First five rows in the dataset**

```python
df.head()
```

**Checking for null values**

```python
df.info()
```

**Shuffling data**

```python
df  = df.sample(frac=1)
```

**Dropping 'CustomerID' feature**

```python
df = df.drop(labels = ['CustomerID'],axis = 1)
```

**Looking for correlations**

```python
corr_matrix = df.corr()
```

**Checking how much age and annual income correlate with the spending score**

```python
corr_matrix['Spending Score (1-100)']
```

**It seems that spending score is higher in young age. Let's visualize this with a scatter plot**

```python
df.plot(kind="scatter", x="Age", y="Spending Score (1-100)")
```

**Handling text feature 'Genre' and normalizing data** 

```python
df.columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder,MinMaxScaler
ct = ColumnTransformer( [("op1", OrdinalEncoder(),['Genre']),("op2", MinMaxScaler(),['Age', 'Annual Income (k$)', 'Spending Score (1-100)'])],remainder = 'passthrough' ) 
                          
df_transform =  ct.fit_transform(df)
```

**Separating inputs and output**

```python
df_transform.shape
```

```python
x = df_transform[:,:3]
y = df_transform[:,-1]
```

**Splitting data in training-set and test-set**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2)
```

**Loading and fitting the model**

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

**Evaluating the model with MSE metric**

```python
from sklearn.metrics import mean_squared_error

preds = model.predict(x_test)
mse = mean_squared_error(y_test,preds)

print('MSE : ',mse)
```
