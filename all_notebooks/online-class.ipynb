# Data file

```python
df= pd.read_csv('/kaggle/input/online-class-prediction/online_class.csv')
df
```

# preprocessing

```python
from sklearn import preprocessing
le= preprocessing.LabelEncoder()

x_1= le.fit_transform(df['sub_code'])
x_2= le.fit_transform(df['day'])
x_3= le.fit_transform(df['time_day'])
x_4= df['attandence']
y_1= le.fit_transform(df['performed'])
```

# for class prediction

```python
x=list(zip(x_1,x_2))
y=y_1
x
```

# for attendance

```python
p1=list(zip(x_1,x_2,y_1))
p2=x_4
```

# KNN classifier

```python
from sklearn.neighbors import KNeighborsClassifier
md= KNeighborsClassifier(n_neighbors=3)
md.fit(x,y)
predict= md.predict([[3,1]])
predict
```

# function

```python
def call(data):
    data_name=['sub_code','day']
    data_x=['x_1','x_2']
    p=0
    en=[]
    gn=[]
    pn=[]
    for j in data:
        for counter, i in enumerate(df[data_name[p]]):
             if i==j:
                # print(j,i,counter)
                 zz=data_x[p]
                 en.append(zz)
                 gn.append(counter)
                 p=p+1
                 break
    pn=list(zip(en,gn))
    #print(pn)
    return pn
```

```python
j=0
for i in range(0,10):
    zn=[]
    pn= call(test_data)
    for i in range(0,2):
        q=pn[i][0]
        r=pn[i][1]
        #print(q,r )
        #print(x[r][i])
        zn.append(x[r][i])
    #print(zn)
    predict1= md.predict([zn])
    if(predict1==1):
        j=j+1
#print(j)
#print(predict1)
```
