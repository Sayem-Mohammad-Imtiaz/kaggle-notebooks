```python
#Reading the data
all_weekly_excess_deaths = pd.read_csv('/kaggle/input/covid19-deaths-dataset/all_weekly_excess_deaths.csv', encoding="windows-1252")
uscounties = pd.read_csv('/kaggle/input/covid19-deaths-dataset/us-counties.csv', encoding="windows-1252")
```

```python
country_list = list(all_weekly_excess_deaths['country'].unique())
country_death_ratio = []
for i in country_list:
    x = all_weekly_excess_deaths[all_weekly_excess_deaths['country']==i]
    country_death_rate = sum(x.covid_deaths)/len(x)
    country_death_ratio.append(country_death_rate)
    
data = pd.DataFrame({'country_list': country_list,'country_death_ratio':country_death_ratio})
new_index = (data['country_death_ratio'].sort_values(ascending=False)).index.values
sorted_data = data.reindex(new_index)

plt.figure(figsize=(15,10))
sns.barplot(x=sorted_data['country_list'], y=sorted_data['country_death_ratio'])
plt.xticks(rotation= 60)
plt.xlabel('Countries')
plt.ylabel('Death Rate')
plt.title('Death Rate Given Countries')
```
