```python
data = pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
# bizim dataset çağrı kodmuz
```

```python

data.rename(columns=({'gender':'censiyet','race/ethnicity':'gruplar'
                     ,'parental level of education':'ebeveyn_egitim_sevyesi','lunch':'besleme','test preparation course':'sinava_hazirlik'
                      ,'math score':'mat_puani','reading score':'okuma_puani'
                     ,'writing score':'yazma_puani'}),inplace=True)

data["censiyet"].replace({"female": "kız", "male": "erkek"}, inplace=True)

data["gruplar"].replace({"group B": "B grubu","group A": "A grubu","group C": "C grubu","group D": "D grubu","group E": "E grubu" }, inplace=True)
data["ebeveyn_egitim_sevyesi"].replace({"bachelor's degree": "lisans derecesi","associate's degree": "ön lisans derecesi",
                                        "high school": "lise derecesi", "master's degree": "yüksek lisans","some college":"universiteli mezun olamadı",
                                        "some high school":"liseli mezun olamadı"
                                       }, inplace=True)

data["besleme"].replace({"standard": "normal", "free/reduced": "düşük"}, inplace=True)
data["sinava_hazirlik"].replace({"none": "yok", "completed": "var"}, inplace=True)

data.info()
```

```python
def harf_puan(marks):
    if marks >= 91:
        return 'AA'
    elif marks >= 82 and marks < 91:
        return 'AB'
    elif marks >=73 and marks < 82:
        return 'BA'
    elif marks >=64 and marks < 73:
        return 'BB'
    elif marks >= 55 and marks < 64:
        return 'CA'
    elif marks >=46 and marks < 55:
        return 'CB'
    elif marks >= 35 and marks < 46:
        return 'CC'
    elif marks < 35:
        return 'FF'
```

```python

data['okuma_puani'] = data['okuma_puani'].apply(harf_puan)
data['yazma_puani'] = data['yazma_puani'].apply(harf_puan)
data['mat_puani'] = data['mat_puani'].apply(harf_puan)
```

```python
def bar_plot(variable):
    
    var = data[variable]
   
    varValue = var.value_counts()
    

    plt.figure(figsize = (25,10))
    plt.bar(varValue.index, varValue)
    plt.xticks(varValue.index, varValue.index.values)
    plt.ylabel("Frequency")
    plt.title(variable)
    plt.show()
    print("{}: \n {}".format(variable,varValue))
```

```python
category1 = ["censiyet","gruplar","ebeveyn_egitim_sevyesi","besleme","sinava_hazirlik", "mat_puani","okuma_puani","yazma_puani"]
for c in category1:
    bar_plot(c)
```
