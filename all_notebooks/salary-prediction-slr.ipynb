```python
data = pd.read_csv('/kaggle/input/salary-data-simple-linear-regression/Salary_Data.csv')
data.head()
```

```python
# shape of Data
data.shape
```

```python
# Numerical Details of data
data.describe()
```

```python
# checking null values in data
data.isnull().sum()
```

```python
# scatter plot b/w dependent and Independent variable
plt.scatter(x=data['YearsExperience'], y=data['Salary'])
plt.show()
```

```python
### As we can see there is a linear relation between the variable. so we can go ahead with our linear regression
```

```python
# creating X and y
X = data['YearsExperience']
y = data['Salary']
```

```python
# importing library for train test split
from sklearn.model_selection import train_test_split

# splitting data into train and test split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=100)
```

```python
# add constant to train data
X_train_sm = sm.add_constant(X_train)
X_test_sm = sm.add_constant(X_test)
```

```python
lr = sm.OLS(y_train, X_train_sm).fit()
lr.params
```

```python
# One benefit of using statsmodel is it gives us statical summary of the model
print(lr.summary())
```

```python
# Plot prediction
plt.scatter(X_train, y_train)
plt.plot(X_train, 25918.438335+9339.081724*X_train, 'r')
plt.show()
```

# **Residual Analysis**

```python
# Residual Analysis
y_train_prediction = lr.predict(X_train_sm)
residual = y_train - y_train_prediction
residual.head()
```

```python
# distribution of residual
sns.distplot(residual)
plt.show()
```

The residuals are following the normally distributed with a mean 0. All good!

```python
# distribution between residual and prediction
plt.scatter(X_train, residual)
plt.show()
```

We are confident that the model fit isn't by chance, and has decent predictive power.

```python
#  Prediction on Test set
y_test_prediction = lr.predict(X_test_sm)
```

```python
# import libraries to see RMSE and R2 score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
train_score = r2_score(y_train, y_train_prediction)
test_score = r2_score(y_test, y_test_prediction)
print('r2 score on train set', train_score)
print('r2 score on test set', test_score)
```

After checking the R2 score on test and train data. Our model is ready and feasible for future predictions
