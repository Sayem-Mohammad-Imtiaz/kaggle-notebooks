##Â Dabl

Enables fast data visualization for features and fits every classifier with their hyperparameters and returns their metrics.

```python
pip install dabl
```

```python
import dabl
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_digits
```

```python
titanic = pd.read_csv(dabl.datasets.data_path("titanic.csv"))
```

```python
titanic.head()
```

```python
titanic_clean = dabl.clean(titanic, verbose=0)
```

```python
titanic_clean.head()
```

```python
types = dabl.detect_types(titanic_clean)
print(types)
```

```python
dabl.plot(titanic, 'survived')
```

```python
fc = dabl.SimpleClassifier(random_state=0)
y = titanic_clean.survived
X = titanic_clean.drop("survived", axis=1)
fc.fit(X, y) 
```
