# 2019 yılı değerleri

**Happiness Score'a göre 2019 verileri incelendiğinde ilk 10 ülke ve Türkiye aşağıda yer almaktadır.**

```python
dt19=alldata[alldata['Year']=='2019']#alldatadan 2019 verisini çek
dt19['Ulke'] = 'Dünya'#Tüm sütuna Dünya verisi gir
dt19.loc[dt19['Country'] == 'Turkey','Ulke'] = 'Türkiye'#Türkiyenin karşısında Türkiye yazsın
```

```python
top10=dt19.nlargest(10,'Happiness Score')
turkey=dt19[dt19['Country']=='Turkey']
top10.drop(['Ulke'], inplace=True, axis=1)
turkey.drop(['Ulke'], inplace=True, axis=1)
top10tr=pd.concat([top10,turkey])
top10trdt=top10tr.set_index('Happiness Rank')
top10trdt
```

**Yukarıda Happiness Score'a göre Top10 ülke ve Türkiye yer almaktadır.**

**Bundan Sonra 2019 verisi üzerinden Happiness Score ile diğer diğer değerlerin ilişkisi ele alınmaktadır. Dünya ülkelerinin yer aldığı scatterplotlarda Türkiye'nin konumu sarı renkle işaretlenmiştir.**

```python
plt.figure(figsize=(14,6))
sns.scatterplot(y=dt19['Happiness Score'], x=dt19['Economy'], hue=dt19['Ulke'],size=dt19['Happiness Score']+dt19['Economy']).set_title("Happiness Score, Economy ilişkisi", fontsize=22)
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(y=dt19['Happiness Score'], x=dt19['Family'], hue=dt19['Ulke'], size=dt19['Happiness Score']+dt19['Family']).set_title("Happiness Score, Family ilişkisi", fontsize=22)
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(y=dt19['Happiness Score'], x=dt19['Life Expectancy'], hue=dt19['Ulke'], size=dt19['Happiness Score']+dt19['Life Expectancy']).set_title("Happiness Score, Life Expectancy ilişkisi", fontsize=22)
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(y=dt19['Happiness Score'], x=dt19['Freedom'], hue=dt19['Ulke'], size=dt19['Happiness Score']+dt19['Freedom']).set_title("Happiness Score, Freedom ilişkisi", fontsize=22)
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(y=dt19['Happiness Score'], x=dt19['Trust'], hue=dt19['Ulke'],size=dt19['Happiness Score']+dt19['Trust']).set_title("Happiness Score, Trust ilişkisi", fontsize=22)
```

```python
plt.figure(figsize=(14,6))
sns.scatterplot(y=dt19['Happiness Score'], x=dt19['Generosity'], hue=dt19['Ulke'],size=dt19['Happiness Score']+dt19['Generosity']).set_title("Happiness Score, Generosity ilişkisi", fontsize=22)
```
