# RSI

***The relative strength index (RSI) is a momentum indicator used in technical analysis that measures the magnitude of recent price changes to evaluate overbought or oversold conditions in the price of a stock or other asset. The RSI is displayed as an oscillator (a line graph that moves between two extremes) and can have a reading from 0 to 100.***

```python
rsi_6 = pd.DataFrame(data1_df['rsi_6'])
rsi_15 = pd.DataFrame(data1_df['rsi_15'])
```

```python
data_df
```

```python
import matplotlib.pyplot as plt

dd = data_df[:50]

plt.figure(figsize=(15,7))
plt.plot(dd['rsi_6'],'g',label="rsi_6")
plt.plot(dd['rsi_15'], 'r', label="rsi_15")
#plt.plot(df["Close"], label="Close")
plt.legend()
plt.show()
```

# Combining to 5min interval

```python
from stock_pandas import StockDataFrame as SDF
data2 = SDF(
    data1_df,
    date_col='open time',
    # Which is equivalent to `time_frame=TimeFrame.M5`
    time_frame='5m'
)
data2
```

```python
data2_15m = data2.cumulate()

data2_15m
```

```python
import matplotlib.pyplot as plt

dd1 = data2_15m[:50]

plt.figure(figsize=(15,7))
plt.plot(dd1['rsi_6'],'g',label="rsi_6")
plt.plot(dd1['rsi_15'], 'r', label="rsi_15")
#plt.plot(df["Close"], label="Close")
plt.legend()
plt.show()
```

# MACD

***Moving average convergence divergence (MACD) is a trend-following momentum indicator that shows the relationship between two moving averages of a securityâ€™s price. ***

```python
macd = data2_15m['macd']
```

```python
data2_15m['macd']= macd
data2_15m
```

```python
import matplotlib.pyplot as plt

dd2 = data2_15m[:300]

plt.figure(figsize=(15,7))
plt.plot(dd2['macd'],'y',label="macd")
#plt.plot(dd2['rsi_15'], 'r', label="rsi_15")
#plt.plot(df["Close"], label="Close")
plt.legend()
plt.show()
```
