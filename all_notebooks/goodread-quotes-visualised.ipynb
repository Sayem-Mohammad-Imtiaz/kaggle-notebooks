```python
quotes = pd.read_csv("/kaggle/input/goodreads-quotes/quotes.csv")
quotes.sample(10)
```

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
```

```python
#quotes
wordcloud = WordCloud(stopwords=STOPWORDS
         ,max_words=100).generate(quotes['quote'].to_string())
plt.figure(figsize=(15,15))
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

```python
#tags
wordcloud = WordCloud(stopwords=STOPWORDS
         ,max_words=100).generate(str(set([y for x in quotes['tags'].str.split(';') for y in x])))
plt.figure(figsize=(15,15))
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

```python
quotes['tags_list']=quotes['tags'].str.split(';') 
qt_ex = quotes.explode('tags_list')
```

```python
# most liked tags
qt_ex.groupby('tags_list')['likes'].sum().nlargest(10).sort_values().plot.barh(colormap='winter',style='k')
```

```python
# most liked author
quotes.groupby('author')['likes'].sum().nlargest(10).sort_values().plot.barh(colormap='winter',style='k')
```

```python
# most liked quote
quotes.groupby('quote')['likes'].sum().nlargest(10).sort_values().plot.barh(colormap='winter',style='k')
```

```python
# author with most number of quotes
quotes.groupby('author')['quote'].count().nlargest(10).sort_values().plot.barh(style='k--')
```
