**COVID-19 : INDIA**

The COVID-19 pandemic in India is part of the worldwide pandemic of coronavirus disease 2019 (COVID-19) caused by severe acute respiratory syndrome coronavirus 2 (SARS-CoV-2). The first case of COVID-19 in India, which originated from China, was reported on 30 January 2020. India currently has the largest number of confirmed cases in Asia, and has the second-highest number of confirmed cases in the world after the United States with more than 10.3 million reported cases of COVID-19 infection and more than 150,000 deaths as of January 06, 2021.The per day cases peaked mid-September in India with over 90,000 cases reported per day and have since come down to below 40,000 in December.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid_df=pd.read_csv('https://raw.githubusercontent.com/Vaibhav15012000/Machine-Learning-Data/main/covid_19_india.csv')
```

```python
covid_df
```

```python
covid_df.head()
```

```python
covid_df.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
plt.plot(covid_df['Confirmed'])
plt.title('Confirmed Cases')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(covid_df['Deaths'])
plt.title('Deaths Cases')
plt.show()
```

```python
covid_df['State/UnionTerritory'].unique()
```

```python
covid_df=covid_df.replace('Telengana','Telangana')
covid_df=covid_df.replace('Telengana***','Telangana')
covid_df=covid_df.replace('Telangana***','Telangana')
covid_df=covid_df.replace('Maharashtra***','Maharashtra')
covid_df=covid_df.replace('Chandigarh***','Chandigarh')
covid_df=covid_df.replace('Punjab***','Punjab')
```

```python
covid_df['State/UnionTerritory'].unique()
```

```python
covid_df.describe()
```

```python
x = covid_df['Confirmed'].sum()
y = covid_df['Cured'].sum()
z= covid_df['Deaths'].sum()
active= x-y
print('Total Confirmed cases =',x)
print('Total Cured cases =',y)
print('Total Active cases =',active)
print('Total Death cases =',z)
```

```python
sns.barplot(x=['Confirmed','Cured','Deaths','active'],y=[x,y,z,active])
```

**> > Top 10 contaminated states**

```python
covid_df1 = covid_df.tail(35)
covid_df2 = covid_df1.sort_values(by='Confirmed', ascending=False).head(10)
```

```python
covid_df2
```

```python
plt.figure(figsize=(20,10))
plt.bar(covid_df2['State/UnionTerritory'],height=covid_df2['Confirmed'],color='yellow')
```

```python
covid_df2['Fatality-Ratio'] = covid_df2['Deaths']/covid_df2['Confirmed']
```

```python
plt.figure(figsize=(20,10))
plt.title('Fatality-Ratio of Top 10 Contaminated States',size=15)
sns.pointplot(data=covid_df2,x='State/UnionTerritory',y='Fatality-Ratio',color='Red')
plt.show()
```

```python
covid_df2['Cure-Ratio'] = covid_df2['Cured']/covid_df2['Confirmed']
```

```python
plt.figure(figsize=(20,10))
plt.title("Cure Ratio")
sns.pointplot(data=covid_df2,x='State/UnionTerritory',y='Cure-Ratio',color='blue')
```

**Now we will visualize the most affected state i.e Maharastra**

```python
mh_df=covid_df[covid_df['State/UnionTerritory']=='Maharashtra']
```

```python
mh_df
```

**Below we have done data filtering as we want a time series data so we have decided a date and filtered that date data only...**

```python
ret_df=mh_df[mh_df['Date'].isin(['09/03/20','09/04/20','09/05/20','09/06/20','09/07/20','09/08/20','09/09/20','09/10/20','09/11/20','09/12/20'])]
```

```python
ret_df
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=ret_df,x='Date',y='Confirmed',color='Blue')
sns.lineplot(data=ret_df,x='Date',y='Cured',color='Green')
sns.lineplot(data=ret_df,x='Date',y='Deaths',color='Red')
plt.show()
```

**Now we will line plot time series of all states data**

```python
ret_df2=covid_df[covid_df['Date'].isin(['09/03/20','09/04/20','09/05/20','09/06/20','09/07/20','09/08/20','09/09/20','09/10/20','09/11/20','09/12/20'])]
```

```python
ret_df2
```

```python
plt.figure(figsize=(20,10))
sns.pointplot(data=ret_df2,x='Date',y='Confirmed',hue='State/UnionTerritory')
plt.legend(bbox_to_anchor=(1,1))
plt.title("Confirmed Cases")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.pointplot(data=ret_df2,x='Date',y='Deaths',hue='State/UnionTerritory')
plt.legend(bbox_to_anchor=(1,1))
plt.title("Death Cases")
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.pointplot(data=ret_df2,x='Date',y='Cured',hue='State/UnionTerritory')
plt.legend(bbox_to_anchor=(1,1))
plt.title("Cured Cases")
plt.show()
```
