```python
import numpy as np
import pandas as pd 
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv("/kaggle/input/dogecoin-historical-data20172021/Dogecoin Historical Data.csv")
df.info()
```

```python
df.head()
```

```python
#remove the '%' and ',' from the change % column
df["Change %"]= df["Change %"].str.replace("%", "")
df["Change %"]= df["Change %"].str.replace(",", "").astype(float)
```

```python
df[df['Price']==df['Price'].min()]
```

```python
df[df['Price']==df['Price'].max()]
```

```python
import matplotlib.pyplot as plt
df.plot()
```

```python
#rConvert the Vol. column to numeric values 
df["Vol."] = df["Vol."].str.replace("B", "*(10**9)")
df["Vol."] = df["Vol."].str.replace("M", "*(10**6)")
df["Vol."] = df["Vol."].str.replace("K", "*(10**3)")
df["Vol."] = df["Vol."].str.replace("-", "0")
df["Vol."] = df["Vol."].str.replace(",", "")
```

```python
#using eval() function to convert into numeric values 
df["Vol."] = df["Vol."].apply(lambda x: eval(x))
```

```python
df.corr()
```

```python
#plotting the co-relation plot
import plotly.graph_objects as go
corr = df.corr()
graph = go.Figure()
graph.add_trace(go.Heatmap(z=corr.values, x=corr.index.values, y=corr.columns.values))
graph.show()
```
