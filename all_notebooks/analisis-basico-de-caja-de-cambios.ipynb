## Gráfica de la serie temporal a1, 1000 puntos

```python
time_serie = dataset.iloc[:,:]
print(time_serie)
```

### Representación gráfica
Señal de los 4 acelerómentros en función del tiempo

```python
# Import Python libraries for graphical representations
import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
rcParams['figure.figsize'] = 15, 8
time_serie.plot()
plt.show()
```

```python
time_serie.plot(style='x')
# plt.show()
```

### Separar las 4 señales en 4 gráficas

```python
a1 = time_serie.iloc[:,0]
a2 = time_serie.iloc[:,1]
a3 = time_serie.iloc[:,2]
a4 = time_serie.iloc[:,3]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

a1.plot(ax=axes[0,0])
a2.plot(ax=axes[0,1])
a3.plot(ax=axes[1,0])
a4.plot(ax=axes[1,1])
```
