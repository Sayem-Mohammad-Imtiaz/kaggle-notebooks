```python
# The following notebook provides the details of finding the accuracy and comparing it with the rest. Here Gaussian Naive Bayes algorithm is used to predict the values.
# As we have used sklearn module, we use GaussianNB function from it. it predicts the test values and compare it with original value and find out the accuracy.
# at first we predict value from Pregnancies to Age:

import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn import model_selection
from sklearn.model_selection import train_test_split

df=pd.read_csv("../input/diabetes-prediction/Diabetespred.csv")
print(df)

x=df.loc[:,'Pregnancies':'Age']
y=df['Outcome']

x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=4,test_size=0.3)

nb=GaussianNB()
nb.fit(x_train,y_train)
pred=nb.predict(x_test)
print(pred)
print(y_test)
print('Accuracy score is {}'.format(nb.score(x_train,y_train)))
```
