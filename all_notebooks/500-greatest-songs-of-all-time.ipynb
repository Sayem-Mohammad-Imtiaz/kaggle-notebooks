```python
song.head()
```

## Data Cleaning

```python
# checking null values if exist
song.isnull().sum()
```

* ### streak and position columns have null values

```python
song.dropna(inplace=True)
```

```python
song.info()
```

```python
# making a new column for year
song['released_year']=song['released'].str.split(', ').str[1]
```

```python
song['released_year'] = song['released_year'].astype(int)
```

```python
# making a new column of int type for streak in weeks
song['streak in weeks']=song['streak'].str[:-5]
```

```python
song['streak in weeks']=pd.to_numeric(song['streak in weeks'],errors='coerce')
```

```python
song.dropna(inplace=True)
```

```python
song['streak in weeks'] = song['streak in weeks'].astype(int)
```

```python
song.info()
```

```python
# deleting streak column
song.drop('streak', axis=1, inplace=True)
```

```python
# making a new column for position number
song['Position']=song['position'].str.split('. ').str[1]
```

```python
song.isnull().sum()
```

```python
song.info()
```

```python
song['Position']=pd.to_numeric(song['Position'],errors='coerce')
```

```python
song.isnull().sum()
```

```python
song.dropna(inplace=True)
```

```python
song['Position'] = song['Position'].astype(int)
```

```python
song.drop('position', axis=1, inplace=True)
```

```python
song.head()
```

```python
# removing duplicates
song.drop_duplicates(inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('dark_background')
```

* ### most number of songs were released in 1965

```python
df=song['released_year'].value_counts().reset_index().rename(columns={'released_year':'count','index':'year'})
df.head()
```
