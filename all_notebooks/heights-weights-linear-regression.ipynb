# Heights & Weights - Linear Regression

## Importing required libraries 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Importing the dataset 

```python
df = pd.read_csv('/kaggle/input/heights-and-weights/data.csv')
x = df.iloc[:, 0]
y = df.iloc[:, 1]
df
```

## Splitting the Dataset into Training Set and Test Set 

```python
from sklearn.model_selection import train_test_split
x = x.values.reshape(-1, 1)
y = y.values
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size = 0.2, random_state = 0)
```

## Training the Simple Linear Regression Model on the Training Set 

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(x_train, y_train)
y_pred = linear_reg.predict(x_test)
```

## Visualisation of the models 

### Training Set Visualization 

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, linear_reg.predict(x_train), color='blue')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.show()
```

### Test Set Visualisation 

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_test, y_pred, color='orange')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.title('Height vs Weight')
plt.show()
```

## Evaluation of the model 

```python
print("Score: ",linear_reg.score(x_test,y_test))
```
