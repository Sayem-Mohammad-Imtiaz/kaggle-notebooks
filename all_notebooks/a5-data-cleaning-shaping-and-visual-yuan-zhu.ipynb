```python
#from this form, I want to just have a subset of columns shown here instead of having to filter through all of them
# I am going to extract the variables that I have more interests in this dataset
#I am going to use subset of "userId" "gender" "age_group""content_views""weekendstrailswatchedperday"""weekdaystrailswatchedperday"
#"avgCompletion" "avgTimeSpent" "avgDuration" "tier" "following_rate" these factors 

#so I am using the subset command in panda
df_subset=df[['userId', 'tier','gender','age_group','weekends_trails_watched_per_day','weekdays_trails_watched_per_day','slot1_trails_watched_per_day','slot2_trails_watched_per_day','slot3_trails_watched_per_day','slot4_trails_watched_per_day','content_views','num_of_comments','avgCompletion','avgTimeSpent','avgDuration','following_rate','avgt2']]
df_subset.head(10)
```

```python
# I want to rank this form by the value of following_rate from high to low
df_subset_frasc=df_subset.sort_values('following_rate',ascending=False)
df_subset_frasc

```

```python
# give me a glance of the summary of this organized form
df_subset_frasc.mean()
 
```

```python
#show a diagram indicating the relationship of "weekdays_trails_watched_per_day" and "age_group"
df_subset_frasc.plot.scatter(x='weekdays_trails_watched_per_day',y='age_group')
df_subset_frasc.plot.scatter(x='weekends_trails_watched_per_day',y='age_group')
#there is some outliers in different age grouops in their weekday's view
df_subset_frasc.plot.scatter(x='weekdays_trails_watched_per_day',y='gender')
df_subset_frasc.plot.scatter(x='weekends_trails_watched_per_day',y='gender')
```
