```python
import pandas as pd
import numpy as np
import datetime
import calendar

import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/kaggle/input/netflix-shows/netflix_titles.csv")
data.head()
```

```python
data.info()
```

```python
#Check for null values 

data.isnull().sum()
```

```python
#Dropping 'na' values from date_added and rating columns

data = data.dropna(subset = ['date_added', 'rating'])
data.shape
```

```python
#Trimming the additional spaces in date column

data['date_added']=data['date_added'].str.strip()
```

```python
#Creating column for formatted date_added column as netflix_date

data['date_added']=data['date_added'].astype(str)
data['netflix_date']=data['date_added'].apply(lambda x: datetime.datetime.strptime(x,'%B %d, %Y').strftime('%Y-%m-%d'))
data.head(2)
```

```python
#Converting 'netflix_date' to date-time obejct, and extracting year and month from the cloumn'

data['netflix_date']=pd.to_datetime(data['netflix_date'], format='%Y-%m-%d')
data['netflix_year']=data['netflix_date'].apply(lambda x:x.year)
data['netflix_month']=data['netflix_date'].apply(lambda x:x.month)
data['netflix_month_abbr']=data['netflix_month'].apply(lambda x:calendar.month_name[x])
data.head(2)
```

```python
print("Recent movie release year in netflix:", data['netflix_year'].max())
print("Old movie release year in netflix:", data['netflix_year'].min())
```

```python
#Splitting the country and listed_in attributes separated by comma and choosing first word 

data['country'] = data['country'].str.split(',', n=1, expand = True)
data['listed_in'] = data['listed_in'].str.split(",", n = 1, expand = True)
data.head(2)
```

```python
#Top 5 directors 

data.director.value_counts().nlargest(5)
```

```python
#Unique values count in Country, Rating and Listed_in columns

data[['country', 'rating', 'listed_in']].nunique()
```

```python
data.pivot_table(index = ['type'], columns = ['rating'], values = ['show_id'], aggfunc = 'count', fill_value = 0)
```

```python
#Movie and TV Show distribution 

fig,ax = plt.subplots(figsize=(5,5))
c = data['type'].value_counts()
lables = c.index
explode = (0, 0.1)
ax.pie(c.values, explode = explode, radius = 1.5, labels = lables, autopct='%1.1f%%')
plt.show()
```

```python
#Rating analysis

sns.set(style='darkgrid')
plt.figure(figsize=(10,7))
sns.countplot(x= 'rating', data = data, order=data['rating'].value_counts().index[0:15])
plt.title("Distribution of ratings across the given years\n")
```

```python
#Release year analysis

plt.figure(figsize=(8,8))
sns.countplot(y='netflix_year', data = data, order = data['netflix_year'].value_counts().index[0:12])
plt.title("New releases in Netflix across years\n")
```

```python
# Number of TVshows and Movies released in each year

plt.figure(figsize=(15,5))
sns.countplot(x = 'netflix_year', hue = 'type', data = data)
plt.title("Number of movies/tvshows released across years\n")
```
