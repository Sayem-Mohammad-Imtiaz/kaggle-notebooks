```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../input/stroke-prediction-dataset/healthcare-dataset-stroke-data.csv").drop("id",axis=1)
data
```

```python
data = data.dropna()
```

```python
X = data.drop("stroke",axis=1)
y = data["stroke"]
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

categorical_features = ["gender","ever_married","work_type","Residence_type","smoking_status"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot",one_hot,categorical_features)],remainder="passthrough")

transformed_X = transformer.fit_transform(X)
transformed_X
```

```python
X = pd.DataFrame(transformed_X)
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X.iloc[:,19:21])
X.iloc[:,19:21] = scaler.transform(X.iloc[:,19:21])
scaler = StandardScaler()
scaler.fit(X.iloc[:,16:17])
X.iloc[:,16:17] = scaler.transform(X.iloc[:,16:17])
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train,y_train)
rf.score(X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression

log = LogisticRegression()
log.fit(X_train,y_train)
log.score(X_test,y_test)
```
