```python
df = pd.read_csv("/kaggle/input/marketing-campaigns-of-portuguese-bank/bank.csv")
df.head(5)
```

```python
bank_df = pd.DataFrame(df)

#Column Name Extraction
single_column = bank_df.columns
total_records = bank_df.shape[0]
columns_name = single_column[0].split(";")
columns_name.remove('age')

column_list = []
column_list.append('age')
for each in columns_name:
    first_quote = each.index("\"")
    second_quote = each.index("\"",1)
    each_name = each[first_quote+1:second_quote]
    #print(each_name)
    column_list.append(each_name)
column_list
```

```python
# Split each record with semi colon
# Extract data only
# Add modified record to Data Frame

each_row = []
for index in range(total_records):
    start = index
    end = index + 1
       
    #print(bank_df[start:end])
    #print("start"+str(start)+" and end "+str(end))
    for each in bank_df.iloc[index]:
        each_record = each.split(";")
        record_list = []
        for each in each_record:
            if(each[0]!="\""):
                each_name = int(each)
            if(each[0]=="\""):
                first_quote = each.index("\"")
                second_quote = each.index("\"",1)
                each_name = each[first_quote+1:second_quote]
            record_list.append(each_name)
    
        
        each_row.append(record_list)
new_df = pd.DataFrame(data = each_row, columns = column_list)
```

```python
#Visualization for each column
for each in column_list:
    frequency = new_df[each].value_counts()
    x = lambda value: new_df[value].unique()
    key = x(each)
    plt.figure(figsize=(14, 5))
    plt.bar(key, frequency,color="r")
    plt.ylabel("Number of Registered People")
    plt.xlabel(each)
    plt.show()
```
