```python
#Libraries

import numpy as np
import pandas as pd
from sklearn.ensemble import IsolationForest
```

```python
#Reading Csv
df=pd.read_csv('../input/anomaly-detection-dataset-for-beginners/Sample scores.csv')
df
```

```python
#Model training
model=IsolationForest(n_estimators=100,max_samples='auto',contamination=float(0.2),max_features=1.0)
model.fit(df[['Scores']])
```

```python
#Prediction
df['anomailes_scores']=model.decision_function(df[['Scores']])
df['anomaly']=model.predict(df[['Scores']])
df.head(20)
```
