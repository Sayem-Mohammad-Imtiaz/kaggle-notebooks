```python
# two main datatypes in pandas
import pandas as pd
```

```python
# 1. Series - one dimensional datatype. It  contains only one single columns. By default index will start from  0
cars = pd.Series(["Audi", 'BMW', 'Range Rover', 'Rolls Royce'])
print(cars)
```

```python
# We can mention index explicit
cars = pd.Series(["Audi", 'BMW', 'Range Rover', 'Rolls Royce'], index = [2,4,6,8])
print(cars)
```

```python
# other way for creating series
colours = ['Purple', 'Black', 'Blue', 'Yellow']  # list is created
colours = pd.Series(colours, index = [2,4,6,8])  # converting list to series format
print(colours)
```

```python
# 2. Data Frame - 2 dimensional datatype. It takes a python Dictionary that contains key and values. We can make df using series as well
car_df = pd.DataFrame({'car_make' : cars, "color" : colours})
print(car_df)
```

```python
# creating dataframe from scratch is a tedious process. so we can import data
# Dataframe contians different datatypes. It contains rows adn columns .
#if axis = 0 it indicates it is a row (by defaults index is 0). If axis = 1, it indicates it is a column (by default index = 1)

df = pd.read_csv("../input/car-sales-dataset/Carsales_dataset.csv")
df.head()
```
