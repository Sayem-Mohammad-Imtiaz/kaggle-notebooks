```python
#To visualize the data
import seaborn as sns
#To work with dataframes
import pandas as pd
#To perform numerical operations
import numpy as np
#To partition the data
from sklearn.model_selection import train_test_split
#importing the library for logistic regression
from sklearn.linear_model import LogisticRegression
#importing performance metrics
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
data_income=pd.read_csv('/kaggle/input/incomejkv/income.csv')
data_income.head()
```

```python
#Input Values
x=data_income.iloc[:,:-1].values
print(data_income.columns)
print("First 5 input features ")
print(x[0:5,:])
```

```python
print("Data Set information")
print(data_income.info)
```

```python
#output values
y=data_income.iloc[:,12].values
print("first five records ")
print(y[0:5])
```

```python
#Data Preprocessing
data=pd.read_csv('/kaggle/input/incomejkv/income.csv',na_values=[" ?"])
#finding null values
print(data.isnull().sum())
#eliminate null values
data1=data.dropna(axis=0)
print(data1.isnull().sum())
```

```python
#remove insignificant variables
cols=["gender","nativecountry","race","JobType"]
new_data=data1.drop(cols,axis=1,inplace=True)
```

```python
#data encoding
cat_salStat={" less than or equal to 50,000":0,
" greater than 50,000":1}
data1["SalStat"].replace(cat_salStat,inplace=True)
#display top 6 records
print(data1["SalStat"].head(6))
#encoding input features
new_data=pd.get_dummies(data1,drop_first=True)
```

```python
#storing the column names
column_list=list(new_data.columns)
print(column_list)
#seperating input features from the data
features=list(set(column_list)-set(["SalStat"]))
print(features)
```

```python
#Storing output variable values in y
y=new_data["SalStat"].values
print(y)
#storing input feature values in x
x=new_data[features].values
print(x)
```

```python
#splitting the data into train and test
train_x,test_x,train_y,test_y=train_test_split(x,y,
                    test_size=0.2,random_state=0)
#creating a instance of logistict regression
logistic=LogisticRegression()
#fitting the values for x and y
logistic.fit(train_x,train_y)
#To display the fitting function attributes such as coef,intercept etc..
print(logistic.coef_)
print(logistic.intercept_)
#prediction from the test data
prediction=logistic.predict(test_x)
print("Logistic Regression 20 classified values:")
print(prediction[0:20])
```

```python
#model evaluation using classification metrics
#confusion metrics - To display correctly classified data 
#and wrongly classified data
confus_matrix=confusion_matrix(test_y,prediction)
print("Confusion matrix")
print(confus_matrix)
#Calculate the accuracy
accu_score=accuracy_score(test_y,prediction)
print("Accuracy")
print(accu_score)
#To display missclassified values from the prediction
print("Missclassified")
print((test_y!=prediction).sum())
```
