```python
import datetime

date = datetime.date.today()
year1 = date.strftime("%Y")
```

```python
#label encoder
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
data2['fuel'] = labelencoder.fit_transform(data2['fuel'])
data2['owner'] = labelencoder.fit_transform(data2['owner'])
```

```python
data2.head()
```

```python
#feature aelection
```

```python
x = data2[['km_driven','fuel','owner','engine','max_power','seats','power','year_old']]
```

```python
y = data2[['selling_price']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state = 10)
```

```python
lnr = LinearRegression()
```

```python
model =lnr.fit(x_train,y_train)
```

```python
predict = model.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,predict)
```

```python
plt.title('Comparison of Y values in test and the Predicted values')
plt.ylabel('Test Set')
plt.xlabel('Predicted values')
plt.scatter( y_test,predict,  color='black')
plt.show()
```
