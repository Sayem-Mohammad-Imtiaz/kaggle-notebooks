# importing the libraries


```python
import plotly.graph_objects as go
from plotly.offline import iplot
```

# Loadind the data

```python
data = pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
```

```python
data.head()
```

```python
data.describe()
```

# analyzing scores by gender

```python
fig = go.Figure()

fig.add_trace(go.Violin(x=data['gender'],
                        y=data['math score'],
                        legendgroup='Math', scalegroup='Math', name='Math'
                        )
             )
fig.add_trace(go.Violin(x=data['gender'],
                        y=data['reading score'],
                        legendgroup='Reading', scalegroup='Reading', name='Reading'
                        )
             )
fig.add_trace(go.Violin(x=data['gender'],
                        y=data['writing score'],
                        legendgroup='Writing', scalegroup='Writing', name='writing'
                        )
             )
fig.update_traces(box_visible=True, meanline_visible=True)
fig.update_layout(violinmode='group',title={
        'text': "Scores by gender",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},  
                  yaxis=dict(
                            title='Scores',
                            titlefont_size=16,
                            tickfont_size=14),
                  xaxis=dict(
                            title='Gender',
                            titlefont_size=16,
                            tickfont_size=14),
                  template = 'plotly_dark'
                  )
iplot(fig)
```

## as can be seen, men had better scores in mathematics and women in reading and writing.
you can always click on the legend to look at each subject individually.
