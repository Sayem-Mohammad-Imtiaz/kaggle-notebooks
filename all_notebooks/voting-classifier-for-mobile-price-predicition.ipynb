# Importing dataset

```python
df = pd.read_csv('/kaggle/input/mobile-price-classification/train.csv')
df
```

```python
x = df.drop(columns=['price_range'])
x
```

```python
y = df['price_range']
y
```

# Train and Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=4)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
knn_clf = KNeighborsClassifier()
gnb_clf = GaussianNB()
rnd_clf = RandomForestClassifier()
dt_clf = DecisionTreeClassifier()
```

# Voting Classifier

```python
voting_clf = VotingClassifier( estimators=[ ('knn',knn_clf),('gnb', gnb_clf),('rf', rnd_clf),('dt',dt_clf)
 ], voting='hard' )
voting_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
for clf in (knn_clf,gnb_clf,rnd_clf,dt_clf,voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

***Pros:***

* Voting classifier is a powerful method and can be a very good option when a single method shows bias towards a particular factor. 
* This method can be used to derive a generalized fit of all the individual models.

***Cons:***

* We should wisely chose the correct model to be used.
* We have to make sure that we do not follow similar mathematical technique.
