```python

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import tensorflow as tf
import tensorflow_hub as hub
```

```python
df = pd.read_csv('../input/appletwittersentimenttexts/apple-twitter-sentiment-texts.csv')
```

```python
train, test = train_test_split(df, test_size=0.20, random_state=42)
```

```python
print(train.shape)
print(test.shape)
```

```python
train['sentiment'].value_counts().plot.pie()
```

```python
test['sentiment'].value_counts().plot.pie()
```

```python
x_train = train['text']
x_test = test['text']
y_train = train['sentiment']
y_test = test['sentiment']
```

```python
x_train = tf.constant(x_train, dtype = tf.string, shape=[1304,])
x_test = tf.constant(x_test, dtype = tf.string, shape=[326,])
y_train = tf.constant(y_train, dtype = tf.float64, shape=[1304,])
y_test = tf.constant(y_test, dtype = tf.float64, shape=[326,])
```

```python
embedding = "https://tfhub.dev/google/universal-sentence-encoder-large/5"
hub_layer = hub.KerasLayer(embedding, input_shape=[], 
                           dtype=tf.string, trainable=True)
```

```python
x_train = hub_layer(x_train)
x_test = hub_layer(x_test)
```

```python
from sklearn.preprocessing import StandardScaler
x_train = StandardScaler().fit_transform(x_train)
x_test = StandardScaler().fit_transform(x_test)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
model = MLPClassifier(random_state=42, verbose=0, early_stopping=True)
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
model.predict(x_test)
```
