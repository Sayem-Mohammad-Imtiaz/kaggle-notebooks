There are two files:

1. full.csv - Dataset that contains complete commit messages of the listed repositories
1. oneline.csv - Top level first line of the commit

The commits are as of April 21, 2021 and contains all the commits of the default branch.

```python
import pandas as pd
```

```python
full_df = pd.read_csv("../input/github-commit-messages-dataset/full.csv")
```

```python
full_df.sample(n=20)
```

```python
oneline_df = pd.read_csv("../input/github-commit-messages-dataset/oneline.csv")
```

```python
oneline_df.sample(n=20)
```

Converting the datatypes

```python
full_df["repo"] = full_df["repo"].astype("category")
full_df["date"] = pd.to_datetime(full_df["date"])
full_df[["commit", "author", "message"]] = full_df[
    ["commit", "author", "message"]
].astype(pd.StringDtype())
```

```python
full_df.dtypes
```

```python
oneline_df[["commit", "message"]] = oneline_df[["commit", "message"]].astype(
    pd.StringDtype()
)
oneline_df["repo"] = oneline_df["repo"].astype("category")
```

```python
oneline_df.dtypes
```

```python
oneline_df.describe()
```
