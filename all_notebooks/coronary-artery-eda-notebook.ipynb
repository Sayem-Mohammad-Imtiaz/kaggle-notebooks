```python
model = Sequential([
    layers.BatchNormalization(),
    layers.Dense(14),
    layers.Activation("relu"),
    layers.Dropout(0.2),
    layers.BatchNormalization(),
    layers.Dense(100),
    layers.Activation("relu"),
    layers.Dropout(0.3),
    layers.BatchNormalization(),
    layers.Dense(100),
    layers.Activation("relu"),
    layers.Dropout(0.2),
    layers.BatchNormalization(),
    layers.Dense(50),
    layers.Activation("relu"),
    layers.Dense(1),
    layers.Activation("softmax"),
], name="CoronaryArteryDisease")
stoping = callbacks.EarlyStopping(min_delta=0.1,patience=10,restore_best_weights=True)
adam = optimizers.Adam(epsilon=0.00001)
model.compile(optimizer=adam,loss="categorical_crossentropy",metrics=["accuracy"])
history=model.fit(trainx,trainy,validation_data=(validx,validy),epochs=75,batch_size=30,callbacks=[stoping])
model.save("model") 
model.save("model.h5")
```
