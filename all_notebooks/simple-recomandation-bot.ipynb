```python
movies_data = pd.read_csv('/kaggle/input/large-movie-dataset/movies_dataset.csv')
movies_data.shape
movies_data['Year']=movies_data['Movie_Name'].str[-5:-1]##geting year from movie name 
movies_data=movies_data[pd.to_numeric(movies_data['Year'], errors='coerce').notnull()]
movies_data['Year'] =movies_data['Year'].astype('int32')
movies_data = movies_data.drop(movies_data[(movies_data['Year'] == 6) | (movies_data['Year'] == 201)].index)
movies_data.shape
```

```python
##Genre = str(input("Genre : ") or "Nolimits") ##-get input from user
Genre = "Nolimits"
```

```python
mov_ye_lim = movies_data.drop(movies_data[~(  (movies_data['Year'] >= init_year) & (movies_data['Year'] <= last_year)    )].index)
print(mov_ye_lim.shape)
if Genre=='Nolimits':
    mov_ye_ge_lim=mov_ye_lim
elif any(mov_ye_lim['Genre'].str.contains(Genre)):
    mov_ye_ge_lim=mov_ye_lim[mov_ye_lim['Genre'].str.contains(Genre)]
else:
    print('wrong selection Genre=0')
    mov_ye_ge_lim=mov_ye_lim
mov_ye_ge_lim.shape

```

```python
semifinal=mov_ye_ge_lim.groupby('Movie_Name')['Rating'].agg(['mean','count'])

fina1=semifinal.sort_values(by=['mean'],ascending=False)
final = fina1.drop(fina1[fina1['count']<L].index)

final.head(N)
```
