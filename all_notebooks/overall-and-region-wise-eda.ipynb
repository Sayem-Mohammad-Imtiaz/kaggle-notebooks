Loading the required libraries and the dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
import matplotlib.colors as mcolors
df = pd.read_csv('../input/videogamesales/vgsales.csv')
```

# Overall Dataset Analysis

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.countplot(x = "Platform", data = df)
```

This shows that most games in the dataset were released on the platforms - DS and PS2.

```python
sns.countplot(x = "Genre", data = df)
```

Most of the games in the dataset are action games.

```python
df1 = df.head(100)
ax = sns.countplot(x="Publisher", data=df1)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

Amongst the top 100 ranked games of all time, more than half of them were published by Nintendo.

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.countplot(x = "Genre", data = df1)
```

Shooter games are the most amongst the top 100, with respect to the Genres.  

#  Analysis of sales in different regions of the world.

# North America

```python
df_na = df.sort_values(by = 'NA_Sales', ascending=False)
```

```python
df_na100 = df_na.head(100)
sns.set(rc={'figure.figsize':(15,10)})
sns.countplot(x = "Genre", data = df_na100)
```

The best selling games in North America are shooter games.

```python
ax = sns.countplot(x="Publisher", data=df_na100)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

The brand with the most games amongst the best selling in North America is Nintendo.

```python
sns.set(rc={'figure.figsize':(15,10)})
sns.countplot(x = "Platform", data = df_na100)
```

This shows that the most famous platform of all time for the North American gamers is the xbox 360.

# Europe

```python
df_eu = df.sort_values(by = 'EU_Sales', ascending=False)
```

```python
df_eu100 = df_eu.head(100)
sns.countplot(x = "Genre", data = df_eu100)
```

The best selling genre of games in Europe is sports games.

```python
ax = sns.countplot(x="Publisher", data=df_eu100)

ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()
```

Just like in North America, Nintendo appears to be the most succesful game publishing brand in Europe.

```python
sns.countplot(x = "Platform", data = df_eu100)
```

The best selling games in Europe were made to be played on the PS3.

# Japan
