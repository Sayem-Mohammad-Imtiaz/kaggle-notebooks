```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

sns.set_style('whitegrid')
plt.style.use("fivethirtyeight")
df=pd.read_csv('../input/covid-thailand-sy/case_1_JUl.csv')

print(df['risk'].value_counts(),20)
```

```python
#clean 
df["sex"]= df["sex"].replace('หญิง','female')
df["sex"]= df["sex"].replace('ชาย','male')
df["province_of_onset"]= df["province_of_onset"].replace('กรุงเทพมหานคร','Bkk')
df["sex"]= df["sex"].replace('หญิง','female')


df["risk"]= df["risk"].replace('สัมผัสใกล้ชิดกับผู้ป่วยยืนยันรายก่อนหน้านี้','close people')
df["risk"]= df["risk"].replace('ทัณฑสถาน/เรือนจำ','Jail')
df["risk"]= df["risk"].replace('อาชีพเสี่ยง เช่น ทำงานในสถานที่แออัด หรือทำงานใกล้ชิดสัมผัสชาวต่างชาติ เป็นต้น','Risk Job')
df["risk"]= df["risk"].replace('คนต่างชาติเดินทางมาจากต่างประเทศ','Aboard')
df["risk"]= df["risk"].replace('คนไทยเดินทางกลับจากต่างประเทศ','Aboard')
df["risk"]= df["risk"].replace('สนามมวย','boxing')
df["risk"]= df["risk"].replace('ไปสถานที่ชุมชน เช่น ตลาดนัด สถานที่ท่องเที่ยว','market')




df.loc[(df['risk'] !='close people') & (df['risk'] !='Jail') & 
       (df['risk'] !='Risk Job' )& (df.risk != 'Aboard')& (df.risk != 'boxing') & (df.risk != 'market'),'risk'] = 'other'

'''
mask = (df.risk != 'close people' and df.risk != 'Jail' and df.risk != 'Risk Job' and df.risk != 'Aboard')
column_name = 'risk'
df.loc[mask, column_name] = 'other'
'''


mask = df.province_of_onset != 'Bkk'
column_name = 'province_of_onset'
df.loc[mask, column_name] = 'other'


mask = df.nationality != 'Thailand'
column_name = 'nationality'
df.loc[mask, column_name] = 'other'

```
