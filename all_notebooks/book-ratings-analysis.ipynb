## Standard Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading Data

```python
data = pd.read_csv('../input/top-270-rated-computer-science-programing-books/prog_book.csv')
```

## Data Cleaning and EDA

```python
data.head()
```

```python
data['Reviews'] = data['Reviews'].apply(lambda x : int(x.replace(',','')))
```

Converting Reviews column into int type

```python
data.describe()
```

```python
plt.figure(figsize=(16,4))
sns.barplot(x = 'Type',y = 'Price',data = data);
```

Looks like **Boxed Set Hardcover** Books are the most expensive (Seems Obvious)

```python
plt.figure(figsize=(16,4))
sns.boxplot(x = 'Type',y = 'Rating',data = data);
```

Rating distribution of different types of Books

```python
sns.distplot(data['Price']);
```

Distribution of Book Prices

```python
sns.jointplot(x = 'Price',y = 'Rating',data = data);
```

Scatter plot of Price with Rating<br>We cannot say much about the Ratings of the book looking at the Price

```python
sns.distplot(data["Reviews"]);
```

```python
sns.heatmap(data.corr(),vmin = -1,vmax = 1,cmap = 'coolwarm');
```
