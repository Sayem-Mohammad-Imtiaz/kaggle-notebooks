```python
from keras.layers import Dense, Flatten 
from keras.models import Sequential
from keras.utils import to_categorical as to_cat
from keras.models import load_model
from keras import models
```

```python
data = pd.read_csv('../input/mnist-in-csv/mnist_train.csv')
```

```python
data.head()
```

```python
X_train = data.iloc[:, 1:].values
X_train=X_train.astype("float32")/255
X_train=X_train.reshape(60000,28,28,1)

 
#X_train.head()
Y_train= data['label'] 
Y_train = to_cat(Y_train, 10) 
#Y_train.head()
#X_train.shape
#Y_train.shape
Y_train
```

```python
testdata = pd.read_csv('../input/mnist-in-csv/mnist_test.csv')
```

```python
testdata.head()
```

```python
X_test = testdata.iloc[:, 1:]. values
X_test=X_test.astype('float32')/255
X_test=X_test.reshape(10000,28,28,1)
#X_test.head()
Y_test= testdata['label'].values
Y_test=to_cat(Y_test, 10) 
#Y_test.head()
#X_test.shape
Y_test.shape
```

```python
model = Sequential() 
model.add(Flatten(input_shape=(28*28,) ))
model.add(Dense(512 ,activation="relu"))
model.add(Dense(10,activation="softmax"))
```

```python
model.compile(optimizer="adam", loss='categorical_crossentropy', metrics=['accuracy'] )
```

```python
model.fit(X_train,Y_train, epochs=10)
```

```python
loss, acc=model.evaluate(X_test, Y_test) 
acc
```

```python
model.save('new_mnist.h5')
```
