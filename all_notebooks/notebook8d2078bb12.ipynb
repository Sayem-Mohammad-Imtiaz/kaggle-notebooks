<h3>Data Cleaning

In this step, we will perform following tasks-
- Make the columns name same for all datasets

- Make the number of columns same

- Check for any null values

- Fill or drop the records with null values

- Change the name of the columns according to their abberevations

- Combine the datasets

- Dropping or adding those columns that are not common for all datasets

- Adding a new column 'Year' to distinguish between data of different years when the dataset is combined

<h4> Since we want to combine the different datasets, we will make sure that the column names is same throughout the dataset.

```python
#Now we will change the name of the columns to their abberevations.
data2015.rename(columns = {'Economy (GDP per Capita)' : 'GDP',
                        'Health (Life Expectancy)' : 'Life',
                        'Trust (Government Corruption)' : 'Trust'}, inplace = True)
```

```python
data2016.rename(columns = {'Economy (GDP per Capita)' : 'GDP',
                        'Health (Life Expectancy)' : 'Life',
                        'Trust (Government Corruption)' : 'Trust'}, inplace = True)
```

```python
data2017.rename(columns={'Happiness.Rank' : 'Happiness Rank',
                        'Happiness.Score' : 'Happiness Score',
                        'Economy..GDP.per.Capita.' : 'GDP',
                        'Health..Life.Expectancy.' : 'Life',
                        'Dystopia.Residual' : 'Dystopia Residual',
                        'Trust..Government.Corruption.' : 'Trust'}, inplace = True)
```

```python

data2018.rename(columns = {'Overall rank' : 'Happiness Rank',
                        'Score' : 'Happiness Score',
                        'Country or region' : 'Country',
                        'Social support' : 'Family',
                        'Freedom to make life choices' : 'Freedom',
                        'GDP per capita' : 'GDP',
                        'Healthy life expectancy' : 'Life',
                        'Perceptions of corruption' : 'Trust'}, inplace = True)
```

<h3> Compare the happiness change across the years

```python
complete_data['Happiness Change'] = (data2018['Happiness Score'] - data2015['Happiness Score']) / data2018['Happiness Score']
```

```python
complete_data.head()
```

```python
complete_data=complete_data.sort_values(by='Happiness Change')
complete_data
```

```python
change=complete_data[complete_data['Happiness Change']>0.00]
change=change.sort_values(by='Happiness Change')
change
```

```python
plt.figure(figsize=(10,7))
plt.title('Changes In The Happiness Index Between 2015 And 2018')
plt.xticks(rotation=90)
sns.barplot(x=change['Country'].tail(10), y=change['Happiness Change'].tail(10),palette='viridis')
```

This shows that Lesotho has showed most positive increase in the Happiness Index Followed by Somaliliand Regionuania, Bhutan etc.

```python
plt.figure(figsize=(10,7))
plt.title('Changes In The Happiness Index Between 2015 And 2018')
plt.xticks(rotation=90)
sns.barplot(x=change['Country'].head(10), y=change['Happiness Change'].head(10),palette='cubehelix')
```

This shows that Italy, despite being a developed nation, has showed least positive increase in the Happiness Index Followed by Lithuania, Beize, etc.

But it must be noted that there are countires with negative change in the index.
