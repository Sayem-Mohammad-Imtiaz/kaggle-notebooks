# Then, we'll read each csv file into a dataframe using pandas, and add each dataframe to the data dictionary using the year (as a string) for the key:

```python
# I've put the first year for you; you'll need to add the rest
data['2007'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2007.csv')

# Rest of your code goes below
data['2008'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2008.csv')
data['2009'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2009.csv')
data['2010'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2010.csv')
data['2011'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2011.csv')
data['2012'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2012.csv')
data['2013'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2013.csv')
data['2014'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2014.csv')
data['2015'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2015.csv')
data['2016'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2016.csv')
data['2017'] = pd.read_csv('/kaggle/input/largest-dog-breed-data-set/2017.csv')
```

# Now we'll visualize the most popular breed from 2017. Here's how many dogs from this breed were licensed in PA from 2007 to 2017:

```python
lab_ret_count = {}
# Store lab retriever counts from each year
for year in data.keys():
    lab_ret_count[year] = data[year].loc[data[year].Breed == "LABRADOR RETRIEVER"].Breed.count()

fig = plt.figure()
ax1 = fig.add_subplot()
ax1.set_xlabel('Year')
ax1.set_ylabel('Count')
ax1.set_title('Lab Retrievers in PA')
ax1.plot(lab_ret_count.keys(),lab_ret_count.values())
```

# Which breed gained the most in popularity from 2008 to 2009?

```python
total_count_09 = data['2009'].Breed.count()
total_count_08 = data['2008'].Breed.count()
breed_counts_09 = data['2009'].Breed.value_counts()
breed_counts_08 = data['2008'].Breed.value_counts()
breed_ratio_09 = breed_counts_09 / total_count_09 * 100
breed_ratio_08 = breed_counts_08 / total_count_08 * 100
diff = (breed_ratio_09 - breed_ratio_08)
diff[diff.notnull()].sort_values(ascending=False)
```

# Let's compare the popularity of Lab Retrievers, Lab Mixes, and German Shepherds

```python
lab_ret_ratio = {}
lab_mix_ratio = {}
ger_shep_ratio = {}
for year in lab_ret_count.keys():
    total = data[year].Breed.count()
    lab_ret_ratio[year] = lab_ret_count[year] / total
    lab_mix_ratio[year] = data[year].Breed.value_counts()['LAB MIX'] / total
    ger_shep_ratio[year] = data[year].Breed.value_counts()['GER SHEPHERD'] / total
    
fig = plt.figure()
ax1 = fig.add_subplot()
ax1.set_title('Popularity of breeds over time')
ax1.set_xlabel('Year')
ax1.set_ylabel('Popularity')
ax1.plot(lab_ret_ratio.keys(),lab_ret_ratio.values())
ax1.plot(lab_mix_ratio.keys(),lab_mix_ratio.values())
ax1.plot(ger_shep_ratio.keys(),ger_shep_ratio.values())
ax1.legend(['Lab Retriever','Lab Mix','German Shepherd'])
```
