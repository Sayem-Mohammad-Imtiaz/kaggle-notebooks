```python
df=pd.read_csv("../input/loan-default-prediction/Default_Fin.csv")
df.head()
df.shape
df.isnull().sum()
df.head()
df['Employed'].value_counts().plot.bar()
df.describe()
df['Annual Salary'].plot.hist()
df.plot.scatter('Bank Balance','Annual Salary')
df.groupby('Employed')['Bank Balance'].mean().plot.bar()
pd.crosstab(df['Defaulted?'],df['Employed'])
df.groupby('Defaulted?')['Bank Balance'].mean().plot.bar()
df.groupby('Defaulted?')['Annual Salary'].mean().plot.bar()
x=df.drop('Defaulted?',axis=1)
y=df['Defaulted?']
df.shape
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=101,stratify=y)
train_y.value_counts()/len(train_y)
test_y.value_counts()/len(test_y)
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
clf.fit(train_x,train_y)
clf.score(train_x,train_y)
clf.score(test_x,test_y)
clf.predict(test_x)
```
