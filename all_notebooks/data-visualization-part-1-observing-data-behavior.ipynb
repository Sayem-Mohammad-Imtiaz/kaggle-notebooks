# **PREDICTING DATA BEHAVIOUR USING BAR PLOTS**
Bar plots are used for discrete DAta and frequencies. From a bar graph, we can see which groups are highest or most common, and how other groups compare against the others.

```python
plt.figure(figsize=(10,5))
plt.bar(iris_df['SepalLengthCm'],height=iris_df['PetalLengthCm'],color='orangered')
plt.title('BAR PLOT(SEPALLENGTH VS PETALLENGTH)',fontsize=15)
plt.xlabel('SEPALLENGTH',fontsize=15)
plt.ylabel('PETALLENGTH',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(iris_df['SepalWidthCm'],height=iris_df['PetalWidthCm'],color='yellow')
plt.title('BAR PLOT(SEPAlWIDTH VS PETALWIDTH)',fontsize=15)
plt.xlabel('SEPALWIDTH',fontsize=15)
plt.ylabel('PETALWIDTH',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(iris_df['SepalLengthCm'],height=iris_df['SepalWidthCm'],color='blue')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length in cms')
plt.ylabel('Sepal Width in cms')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(iris_df['PetalLengthCm'],height=iris_df['PetalWidthCm'],color='purple')
plt.title('Petal Length vs Petal Width')
plt.xlabel('petal Length')
plt.ylabel('Petal Width')
plt.show()
```

# **PREDICTING DATA BEHAVIOUR USING HISTOGRAM PLOTS-FOR CONTINUOUS DATA DISTRIBUTUTION**

A histogram is the most commonly used graph to show frequency distributions. It looks very much like a bar chart, but there are important differences between them.
A frequency distribution shows how often each different value in a set of data occurs

```python
plt.figure(figsize=(10,5))
plt.hist(iris_df['SepalLengthCm'],color='purple',bins=25)   # here bins represent no of divisons in histogram
plt.title('******Histogram*****',fontsize=15)
plt.xlabel('Row indexes',fontsize=15)
plt.ylabel('Sepal Length',fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(iris_df['SepalWidthCm'],color='orangered',bins=20)
plt.title('Histogram Plot')
plt.xlabel('Row Indexes')
plt.ylabel('Sepal Width')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(iris_df['PetalWidthCm'],color='cyan')
plt.title('Histogram Plot')
plt.xlabel('Row Indexes')
plt.ylabel('Petal Width')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(iris_df['PetalLengthCm'],color='purple')  
plt.title('******Histogram*****',fontsize=15)
plt.xlabel('Row indexes',fontsize=15)
plt.ylabel('Petal Length',fontsize=15)
plt.show()
```

# **PREDICTING DATA BEHAVIOUR USING SCATTER PLOTS-relationship plot between X and Y**

Scatter plots' primary uses are to observe and show relationships between two numeric variables. The dots in a scatter plot not only report the values of individual data points, but also patterns when the data are taken as a whole. ... A scatter plot can also be useful for identifying other patterns in data



```python
plt.figure(figsize=(10,5))
plt.scatter(x=iris_df['SepalLengthCm'],y=iris_df['SepalLengthCm'],marker='*',color='red',label='Sepal Width' )
plt.scatter(x=iris_df['SepalLengthCm'],y=iris_df['PetalWidthCm'],marker='s',color='blue',label='Petal Width')
plt.title('SCATTER PLOT',fontsize=15)
plt.xlabel('Sepal Length',fontsize=15)
plt.legend(loc=4)   #for showing labels data and 'loc' atrribute is used for position

plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='SepalLengthCm',y='SepalWidthCm',data=iris_df,marker='s', hue='Species')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x='PetalLengthCm',y='PetalWidthCm',data=iris_df,marker='s',hue='Species',palette='copper')
```

# **Predicting Data Behaviour using Box Plots-used to detect outliers**

A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary. ... It is often used in explanatory data analysis. This type of graph is used to show the shape of the distribution, its central value, and its variability.

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=iris_df)
plt.show()
```

```python
plt.figure(figsize=(10,5))
#orient is used for orientation 'h' means horizontal and by default it is vertical
sns.boxplot(data=iris_df,orient='h',palette='magma')

plt.show()
```
