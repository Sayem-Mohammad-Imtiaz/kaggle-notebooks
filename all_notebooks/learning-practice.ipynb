**WHILE ** and **FOR LOOPS**

```python
# Stay in loop if condition( i is not equal 5) is true. 
i= 0
while i != 5: 
    print("i is :" ,i)
    i+=1
print (i, "is the last")
```

```python
# Stay in loop if i is in the list
lis=[1,2,3,4,5]
for i in lis:
    print ("i is ", i)
print ("Thats it!")
```

```python
# we use enumerate for lists
for index, value in enumerate(lis):
    print(index, ":" , value)    
```

>  **SCOPE**
> > 
> > The scope of a variable in python is that part of the code where it is visible.
> > There are 4 types of Variable Scope in Python (Local,global,enclosing and built-in)
> > Built-in Scope is the widest scope. 
> > The built-in scope has all the names that are loaded into python variable scope when we start the interpreter. For example, we never need to import any module to access functions like print() and id().

```python
x=2 #Global variable (Global scope)
def f():
    x=3  # local variable(local scope)
    return x
print(x)
print(f())
#if there is no local variable, It will take the global one
def g():
    return x
print (g())
```

```python
#LAMBDA FUNCTION
#Faster way of writing function

square = lambda x: x**2
print (square(2))

tot= lambda x,y,z : x+y+z
print(tot(2,3,4))
```

```python
#ANONYMOUS FUNCTION-The difference is it can take more than one variable

alist = [1,2,3]
x = map(lambda x: x**2, alist) #gives the value for x one by one by usşng each elements of the alist
print(list(x))  #we need to show all the x values in a list 
```

```python
#zip it doesnt work with strings but only integers. Why?
list1 = [1,2,3,4]
list2 = [5,6,7,8]
zip12 = zip(list1, list2)
print(zip12)
zip_list = list(zip12)
print(zip_list)

#unzip

unzip = zip(*zip_list)
unlist1, unlist2 = list(unzip) #unzip returns tuble. Converted to list. 
print(unlist1)
print(unlist2)
print (type(unlist1))
```

> **LIST COMPREHENSİON**
> 
>  > creating a list based on existing lists

>  > provide a concise way to create lists

>  > always returns a result list

>  > the result will be a new list resulting from evaluating the expression in the
context of the for and if clauses

>  > the list comprehension starts with a '[' and ']', to help you remember that the
result is going to be a list.

```python
list1 = [1,2,3]
list2 = [2*i for i in list1]  #list of comprehension
print(list2)
```

```python
#Let's see how we use list comprehension in data analysis
#Lets find the speeds and label them high or low according to their posion regarding to avarage speed
threshold = sum(data.Speed) / len(data.Speed)
print('Threshold is: ', threshold)
data['speed_level'] = ['high' if i >threshold else 'low' for i in data.Speed]
data.loc[:10,['speed_level','Speed']]
```
