```python
plt.figure(figsize=(12, 6))
sns.regplot(x='Ladder score', y='Healthy life expectancy', data=data2021, ci=None, color='blue');
plt.title('Correlación entre el puntaje de felicidad y la esperanza de vida', fontsize=20, fontweight='bold', pad=20);
```

En este caso, el puntaje de felicidad y la esperanza de vida se encuentran fuertemente relacionados, en los países que tienen un puntaje más alto de felicidad también es muy notorio que hay una mayor esperanza de vida que en los demás países con puntajes más bajos de felicidad

```python
plt.figure(figsize=(12, 6))
sns.regplot(x='Ladder score', y='Social support', data=data2021, ci=None, color='blue');
plt.title('Correlación entre el puntaje de felicidad y la asistencia social', fontsize=20, fontweight='bold', pad=20);
```

En este caso, el puntaje de felicidad y la asistencia social se encuentran fuertemente relacionados, en los países que tienen un puntaje más alto de felicidad también es muy notorio que existe una mayor asistencia social que en los demás países con puntajes más bajos de felicidad

```python
plt.figure(figsize=(12, 6))
sns.regplot(x='Ladder score', y='Generosity', data=data2021, ci=None, color='blue');
plt.title('Correlación entre el puntaje de felicidad y la generosidad', fontsize=20, fontweight='bold', pad=20);
```

En este caso, el puntaje de felicidad y la generosidad tienen una baja correlación, independientemente del puntaje de generosidad de cada país, no hay una gran diferencia entre el puntaje de felicidad

```python
plt.figure(figsize=(12, 6))
sns.regplot(x='Ladder score', y='Logged GDP per capita', data=data2021, ci=None, color='blue');
plt.title('Correlación entre el puntaje de felicidad y el PIB', fontsize=20, fontweight='bold', pad=20);
```

En este caso, el puntaje de felicidad y el PIB se encuentran fuertemente relacionados, en los países que tienen un puntaje más alto de felicidad también es muy notorio que hay un mayor PIB que en los demás países con puntajes más bajos de felicidad

```python
plt.figure(figsize=(12, 6))
sns.regplot(x='Ladder score', y='Perceptions of corruption', data=data2021, ci=None, color='blue');
plt.title('Correlación entre el puntaje de felicidad y la percepción de corrupción', fontsize=20, fontweight='bold', pad=20);
```

En la gráfica anterior se puede observar que el puntaje de felicidad y la percepción de corrupción están ligeramente relacionados, en los países que tienen un puntaje más alto de felicidad, la percepción de corrupción es ligeramente más baja que en los demás países
