```python
data.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
#we are dropping it because it will not impact the predictions 
```

```python
geography=pd.get_dummies(data['Geography'],drop_first=True)
gender=pd.get_dummies(data['Gender'],drop_first=True)
```

```python
data.drop(['Gender','Geography'],axis=1,inplace=True)
```

```python
data
```

```python
data=pd.concat([data,gender,geography],axis=1)
```

```python
data
```

```python
data.columns
```

```python
X=data[['CreditScore', 'Age', 'Tenure', 'Balance', 'NumOfProducts', 'HasCrCard',
       'IsActiveMember', 'EstimatedSalary','Male', 'Germany',
       'Spain']]
X=X.values
X
```

```python
y=data['Exited']
y=y.values
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss')
```

```python
model= Sequential()
model.add(Dense(units=11,activation='relu',kernel_initializer='he_uniform'))

model.add(Dense(units=7,activation='relu',kernel_initializer='he_uniform'))
model.add(Dense(units=4,activation='relu',kernel_initializer='he_uniform'))
model.add(Dense(units=1,activation='sigmoid',kernel_initializer="glorot_uniform"))
model.compile(optimizer='adam',loss="binary_crossentropy",metrics=['accuracy'])

model.fit(X_train,y_train,batch_size=10,validation_split=0.33,epochs=300,validation_data=(X_test,y_test),callbacks=[early_stop])
```

```python
history=pd.DataFrame(model.history.history)
history
```

```python
history[['loss','val_loss']].plot()
```

```python
history['accuracy'].plot()
```

```python
y_pred = model.predict_classes(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```
