As we have similar data in all the different csv files, we will concatenate together in dataframe using for loop

```python
import glob
path =r'../input' # use your path
allFiles = glob.glob(path + "/*.csv")
frame = pd.DataFrame()
list_ = []
for file_ in allFiles:
    df = pd.read_csv(file_,index_col=None)
    list_.append(df)
pumpkin_data = pd.concat(list_)
```

```python
# full dataframe info
print(pumpkin_data.info())
```

```python
print(pumpkin_data.head())
```

```python
print(pumpkin_data.columns)
```

From info, there are lot of columns which have no values so we just drop those columns

```python
pumpkin_filtered_data = pumpkin_data.drop(['Appearance','Comments','Condition','Crop','Environment','Grade','Offerings','Price Comment','Quality','Storage','Trans Mode','Commodity Name'], axis =1)
```

```python
# filtered dataframe info
pumpkin_filtered_data.info()
```

Question 1 : Where are pumpkin prices highest ?

```python
# firstly find the maximum high price
pumpkin_filtered_data['High Price'].max()
```

```python
# retrieve all the data for maximum high price
pumpkin_filtered_data[pumpkin_filtered_data['High Price'] == 480.0]
```

Question 2 : Where are pumpkin prices lowest ?

Question 3 : How does pumpkin size relate to price?

```python
pumpkin_filtered_data.groupby(['Item Size'])['High Price'].mean().plot(kind='bar')
plt.ylabel("Mean High Pumpkin Price")
```

The plot clearly suggests small pumpkins will fetch low price 

```python
pumpkin_filtered_data['Item Size'].value_counts()
```

Since small pumpkins have the maximum count so we check the distribution of price according to size using median

```python
pumpkin_filtered_data.groupby(['Item Size'])['High Price'].median().plot(kind='bar')
plt.ylabel("Median High Pumpkin Price")
```

Using median clearly depicts size has strong influence on price

Question 4 : Are there any other features of pumpkin relate to price?

```python
# checking color of pumpkin
pumpkin_filtered_data.groupby(['Color'])['High Price'].mean().plot(kind='bar')
plt.ylabel("Mean High Pumpkin Price")
```

Clearly Orange pumpkins fetch maximum price

```python
pumpkin_filtered_data['Color'].value_counts()
```

Clearly Orange color pumpkin are maximum in the market, let's study median of price as well in relation to color

```python
pumpkin_filtered_data.groupby(['Color'])['High Price'].median().plot(kind='bar')
plt.ylabel("Median High Pumpkin Price")
```

Another feature we would like to study is how price varies with packaging.

```python
pumpkin_filtered_data.groupby(['Package'])['High Price'].mean().plot(kind='bar')
plt.ylabel("Mean High Pumpkin Price")
```

Here different packing are available but most interesting is the pumpkin sold as each so let's look how high is the each price according to Item size

```python
pumpkin_filtered_data[pumpkin_filtered_data.Package == 'each'].groupby('Item Size')['High Price'].max()
```

This table also shows small pumpkins sold for very less price. 
Question 5 : Which city sell maximum number of pumpkin varieties ?

```python
pumpkin_filtered_data.dropna(axis=0, how='all').groupby(['City Name'])['Variety'].nunique().plot(kind='bar')
plt.ylabel("Number of Varieties")
```

The answer is Columbia.

Question 6 : Are there any varieties sold by all cities ?

Question 7 : Which city has maximum average high price across all cities ?

Boston has max. average high price. One thing to consider here although San Francisco has highest high price but it's average high price is much lower compared to max. average high price.
