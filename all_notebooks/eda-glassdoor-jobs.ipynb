```python
data[['job_simp', 'Rating', 'min_salary', 'max_salary', 'avg_salary', 'same_state', 'Industry', 'Sector', 'same_state', 'company_age', 'python', 'excel', 'hadoop', 'spark', 'aws', 'tableau', 'big_data']].describe(include = 'all')
```

```python
data['Industry'].unique()
```

```python
ax = data['Industry'].value_counts().plot(kind = 'barh', figsize = (20, 26), title = 'Industry Count', color = 'b')
ax.set_xlabel('Count')
ax.set_ylabel('Industries')
```

```python
data['job_simp'].value_counts().plot(kind = 'barh', figsize  = (16, 8), title = 'Job Categories', color = 'g')
ax.set_xlabel = ('Count')
ax.set_ylabel = ('Job Categories')
```

```python
pd.pivot_table(data, index = 'job_simp', values = 'avg_salary').sort_values(by = 'avg_salary')
```

```python
pd.set_option('display.max_rows', None)
a = pd.pivot_table(data, index = ['job_state', 'job_simp'], values = 'avg_salary').sort_values(by = 'avg_salary', ascending = False)
a
```

```python
fig, ax = plt.subplots(figsize = (26, 26))
sns.barplot(x = 'job_simp', y = 'avg_salary', data = data)
```
