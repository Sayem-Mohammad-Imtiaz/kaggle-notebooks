* Gender: Gender of the passengers (Female, Male)

* 機内wifiサービス：機内wifiサービスの満足度（0：該当なし、1〜5）
* Departure/Arrival time convenient: Satisfaction level of Departure/Arrival time convenient　
* 便利な出発/到着時間：便利な出発/到着時間の満足度
* Ease of Online booking: Satisfaction level of online booking　
* オンライン予約のしやすさ：オンライン予約の満足度
* Gate location: Satisfaction level of Gate location　
* ゲート位置：ゲート位置の満足度
* Food and drink: Satisfaction level of Food and drink　
* 飲食物：飲食物の満足度
* Online boarding: Satisfaction level of online boarding　
* オンライン搭乗：オンライン搭乗の満足度
* Seat comfort: Satisfaction level of Seat comfort　
* 座席の快適さ：座席の快適さの満足度
* Inflight entertainment: Satisfaction level of inflight entertainment　
* 機内エンターテインメント：機内エンターテインメントの満足度
* On-board service: Satisfaction level of On-board service　
* 船内サービス：船内サービスの満足度
* Leg room service: Satisfaction level of Leg room service　
* レッグルームサービス：レッグルームサービスの満足度
* Baggage handling: Satisfaction level of baggage handling　
* 手荷物取り扱い：手荷物取り扱いの満足度
* Check-in service: Satisfaction level of Check-in service　
* チェックインサービス：チェックインサービスの満足度
* Inflight service: Satisfaction level of inflight service　
* 機内サービス：機内サービスの満足度
* Cleanliness: Satisfaction level of Cleanliness　
* 清潔度：清潔度の満足度
* Departure Delay in Minutes: Minutes delayed when departure　
* 分単位の出発遅延：出発時の分遅延
* Arrival Delay in Minutes: Minutes delayed when Arrival　
* 分で到着遅延：議事録は、いつ到着遅れ

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

X_list = ['Age', 'Flight Distance', 'Inflight wifi service',
       'Departure/Arrival time convenient', 'Ease of Online booking',
       'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort',
       'Inflight entertainment', 'On-board service', 'Leg room service',
       'Baggage handling', 'Checkin service', 'Inflight service',
       'Cleanliness', 'Departure Delay in Minutes', 'Gender_Female',
       'Customer Type_Loyal Customer', 'Type of Travel_Business travel',
       'Class_Business']
y_list = ['satisfaction_satisfied']

X = df_tr[X_list]
y = df_tr[y_list]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
logit = LogisticRegression()
logit.fit(X_train, y_train)
```

```python
logit.coef_
```

```python
logit.intercept_
```

```python
y_pred = logit.predict(X_test)
```

```python
y_pred
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier(max_depth=12)
```

```python
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf_r = RandomForestClassifier(n_estimators=10, max_depth=12)
```

```python
clf_r.fit(X_train, y_train)
```

```python
y_pred = clf_r.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
importances = clf_r.feature_importances_
importances
```

```python
features = np.array(df_tr.columns)
```
