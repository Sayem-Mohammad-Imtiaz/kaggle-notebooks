```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
%matplotlib inline

```

```python
df = pd.read_csv('../input/clsuterpenjualan/barang_keluar_imam.csv', delimiter=';', skiprows=0, low_memory=False)

df.head()
```

```python
plt.scatter(df['jumlah_transaksi'],df['total_penjualan'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[['jumlah_transaksi','total_penjualan']])
y_predicted
```

```python
df['cluster'] = y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
# df4 = df[df.cluster==3]

plt.scatter(df1.jumlah_transaksi,df1['total_penjualan'],color='green')
plt.scatter(df2.jumlah_transaksi,df2['total_penjualan'],color='red')
plt.scatter(df3.jumlah_transaksi,df3['total_penjualan'],color='blue')
# plt.scatter(df4.jumlah_transaksi,df4['total_penjualan'],df4['rata_rata'],color='gray')

plt.xlabel('Jumlah Transaksi')
plt.ylabel('Total Penjualan')
plt.legend()
```

```python
scaler = MinMaxScaler()
scaler.fit(df[['total_penjualan']])
df['total_penjualan']= scaler.transform(df[['total_penjualan']])

scaler.fit(df[['jumlah_transaksi']])
df['jumlah_transaksi'] = scaler.transform(df[['jumlah_transaksi']])

# scaler.fit(df['rata_rata'])
# df['rata_rata'] = scaler.transform(df['rata_rata'])
# df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[['jumlah_transaksi','total_penjualan']])
y_predicted
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
# df4 = df[df.cluster==3]

plt.scatter(df1.jumlah_transaksi,df1['total_penjualan'],color='green',label='cluster 0')
plt.scatter(df2.jumlah_transaksi,df2['total_penjualan'],color='red',label='cluster 1')
plt.scatter(df3.jumlah_transaksi,df3['total_penjualan'],color='blue',label='cluster 2')
# plt.scatter(df4.jumlah_transaksi,df4['total_penjualan'],df4['rata_rata'],color='gray')


plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='black',marker='*',label='centroid' )

plt.legend()
```

```python
k_rng = range(1,10)
sse = []

for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[['jumlah_transaksi','total_penjualan']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of Squared Error')
plt.plot(k_rng, sse)
```

```python
selected_cols = ["jumlah_transaksi","total_penjualan"]
cluster_data = df.loc[:,selected_cols]

kmeans_sel = KMeans(init='k-means++', n_clusters=3, n_init=100, random_state=2).fit(cluster_data)
labels = pd.DataFrame(kmeans_sel.labels_)
clustered_data = cluster_data.assign(Cluster=labels)
```

```python
grouped_km = clustered_data.groupby(['Cluster']).mean().round(1)
grouped_km
```
