```python
diff_reasons=['Property Disputes','Family Dispute','Petty Quarrels','Money Disputes','Personal Vendetta','Love Affairs','Casteism','Unknown reasons']
d={0:0,'Murder Happened with reason':1,'Property Disputes':2,'Family Dispute':3,'Petty Quarrels':4,'Money Disputes':5,'Personal Vendetta':6,'Love Affairs':7,'Casteism':8,'Unknown reasons':9}
type(diff_reasons)
```

```python
df[1]=df[1].apply(lambda x : d[x])
```

```python
df.head(25)
```

```python
df[11]=0
df[12]=0
```

```python
multiple_crimes=[]
for i in range(len(df)):
    df.iloc[i,8]=(str(df.iloc[i,8])).replace(" ","")
    #print(df.iloc[i,8])
    if 4>=len(str(df.iloc[i,8]))>=3:
        a,b=df.iloc[i,8].split(sep=",",maxsplit=2)
        df.iloc[i,8]=a
        df.iloc[i,11]=b
        df.iloc[i,12]=0
        multiple_crimes.append(i)
    if len(str(df.iloc[i,8]))>=5:
        df.iloc[i,8],df.iloc[i,11],df.iloc[i,12]=df.iloc[i,8].split(sep=",",maxsplit=3)
        multiple_crimes.append(i)
```

```python
crime_type=df[1].value_counts()
crime_type=crime_type.drop([0])
```

```python
crime_type=crime_type.sort_index()
crime_type.index=diff_reasons
print(crime_type)
```

```python
xaxis=crime_type.index
print(xaxis)
```

```python
yaxis=crime_type.array
```

```python
plt.bar(xaxis,yaxis)
plt.title('different Reasons of crime')
plt.xticks(rotation=90)
plt.ylabel('Number of crimes')

```

```python
df[13]=df[2]+df[3]+df[4]#total victims
```

```python
total_vic=df.groupby(by=0).sum()
print(total_vic)
```

```python
total_crimes=df.groupby(by=0).count()#total crimes
```

```python
city_names=total_crimes.index
#print(city_names)
tyaxis=total_crimes[2]
plt.bar(city_names,tyaxis,color='orange')
plt.title('No of crimes city wise')
plt.xlabel('city')
plt.ylabel('no of crimes')
```

```python
#print(total_vic)
vic_y_axis=total_vic[13]
#print(vic_y_axis)
plt.bar(city_names,vic_y_axis,color='green')
plt.title(' victims citywise')
plt.xlabel('city')
plt.ylabel('no of victims')
```

```python
crime_to_victim=[]
for i in range(len(vic_y_axis)):
    k=vic_y_axis[i]/tyaxis[i]
    crime_to_victim.append(k)

plt.scatter(city_names,crime_to_victim)
plt.title('Crime to victim ratio')
print('An interesting data here as clearly though kanpur is last in both crime rate and victims but the ratio of crimes that result in victims is quite high in kanpur ')        
```
