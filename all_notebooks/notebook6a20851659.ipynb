```python
netflix.info()
```

# Data cleaning

```python
netflix.isnull().sum().sort_values(ascending=False)
```

```python
netflix.drop(['director','cast'],inplace=True,axis=1)
```

```python
netflix['rating']=netflix['rating'].fillna(netflix['rating'].mode()[0])
```

```python
netflix['country']=netflix['country'].fillna(netflix['country'].mode()[0])
```

```python
netflix.isnull().sum().sort_values(ascending=False)
```

```python
netflix['date_added']=netflix['date_added'].fillna('January 1 ,{}'.format(str(netflix['release_year'].mode()[0])))
```

```python
netflix.info()
```

# Data Visualization
Below we will see various graphs to count and show the relation between different features

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

count of movies vs Tv show

```python
plt.figure(figsize=(8,6))
ax=sns.countplot(x='type',data=netflix,palette='Set2')
plt.title('Movies Vs Shows Count')
```

```python
plt.figure(figsize=(8,6))
ax=sns.countplot(x='rating',data=netflix)
plt.title('Rating of Movies and Shows')
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='rating',hue='type',data=netflix,palette='Set2')
plt.title('Rating of Movies Vs TV Shows')
plt.xticks(rotation=90)
plt.xlabel('Rating')
```

```python
plt.figure(figsize=(8,6))
ax=sns.countplot(x='country',data=netflix,order=netflix['country'].value_counts().index[0:15])
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('No of Movies and Shows')
```

```python
plt.figure(figsize=(8,6))
ax=sns.countplot(x='country',data=netflix,hue='type',order=netflix['country'].value_counts().index[0:15])
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('No of Movies Vs Shows')
plt.title('Most Number of Movies Vs Shows top 15 Countries')
```

```python
plt.figure(figsize=(8,6))
ax=sns.countplot(x='country',data=netflix,hue='type',order=netflix['country'].value_counts().index[0:2],palette='Set2')
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('No of Movies Vs Shows')
plt.title('Movies Vs Shows count in Top two countries')
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x='release_year',data=netflix,order=netflix['release_year'].value_counts().index[0:15])
plt.title('Most movies and shows Released in a year')
plt.xlabel('Year of release')
plt.xticks(rotation=90)
```

 #  Count of Movies Vs Shows released in last 15 years

```python
plt.figure(figsize=(10,8))
sns.countplot(x='release_year',data=netflix,hue='type',palette='Set2',order=netflix['release_year'].value_counts().index[0:15])
plt.title('Movies vs Shows released in last 15 years')
plt.xlabel('Year of release')
plt.xticks(rotation=90)
```
