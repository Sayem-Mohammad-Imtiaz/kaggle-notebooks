```python
#Support Vector Regression

import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
from matplotlib.ticker import FuncFormatter

import warnings
warnings.filterwarnings("ignore")

#import dataset
dataset = pd.read_csv('/kaggle/input/position-salaries/Position_Salaries.csv')

dataset
```

```python
#Create matrix of independent variables
X = dataset.iloc[:,1:2].values #[:,1] would return an array but we want X to be a 
#matrix

#Create array of dependent variable
y = dataset.iloc[:,2].values
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
sc_y = StandardScaler()
X2 = X #save untransformed X
X = sc_X.fit_transform(X)
y = sc_y.fit_transform(y.reshape(-1,1))
```

```python
#Fitting SVR to the dataset
from sklearn.svm import SVR
regressor = SVR(kernel='rbf') #rbf = gaussian
regressor.fit(X,y)
```

```python
#Predict value for level 6.5
y_pred = sc_y.inverse_transform(regressor.predict(sc_X.transform(np.array([[6.5]]))))
y_pred
```

```python
def conv_sal(x, pos):
    'The two args are the value and tick position'
    return '$%.0f' % sc_y.inverse_transform(np.array([[x]]))
```

```python
def conv_lev(x, pos):
    'The two args are the value and tick position'
    return '$%1.1f' % sc_X.inverse_transform(np.array([[x]]))
```

```python
#Visualize the SVR results - Scaled Axes
plt.scatter(X, y, color='red')
plt.plot(X, regressor.predict(X), color='blue')
plt.title("Support Vector Regression - Gaussian Kernel")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

```python
#Visualize the SVR results - Scaled Axes
plt.scatter(X, y, color='red')
plt.plot(X, regressor.predict(X), color='blue')
plt.title("Support Vector Regression - Gaussian Kernel")
plt.xlabel("Position Level")
plt.ylabel("Salary")
formatter = FuncFormatter(conv_sal)
plt.gca().yaxis.set_major_formatter(formatter)
locs, labels = plt.yticks()
plt.show()
```

```python
#Visualize the SVR results - Scaled Axes

#since inverse transform of the default ticks yields levels
#that are not whole numbers, we will force the tick values
tix = list(range(11))
tix2 = list(range(11))
label = list(range(11))
label2 = list(range(11))

for i in range(11):
    tix[i] = sc_y.transform(np.array([[i*100000]]))
    tix2[i] = sc_X.transform(np.array([[i]]))
    label[i] = i*100000
    label2[i] = i

plt.scatter(X, y, color='red')
plt.plot(X, regressor.predict(X), color='blue')
plt.title("Support Vector Regression - Gaussian Kernel")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.yticks(tix, label)
plt.xticks(tix2, label2)
plt.xlim(sc_X.transform(np.array([[0.5]])),sc_X.transform(np.array([[10.5]])))
plt.show()
```
