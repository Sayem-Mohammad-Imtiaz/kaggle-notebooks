**part1: preparatory exercises**

```python
#rename columns
df = pd.read_csv("../input/telco-customer-churn/WA_Fn-UseC_-Telco-Customer-Churn.csv" )
df.head()
```

```python
ndf = df.rename(columns={"tenure":'months'})
ndf.head(5)
```

```python
#change casing
# Suppress trailing spaces
ndf["InternetService"].str.strip()
```

```python
#normalize data types
print(ndf)
ndf.dtypes
```

```python
#ndf['TotalCharges'].type: object  --- float
# 将TotalCharges列中的空值全部替换为nan
ndf['TotalCharges'] = ndf['TotalCharges'].replace(' ', np.nan)

# 将值转变为浮点型
ndf['TotalCharges'] = ndf['TotalCharges'].astype(float)
print(ndf['TotalCharges'])
```
