# The concept of training:
* the **last version (commit)** of the notebook has:
        * the basic tasks (after "TASK:")
        * the additional tasks for self-execution (after "ADDITIONAL TASK:")
* the **previuos version (commit)** of the notebook has **answers** for the basic tasks

### The title of each notebook of the course contains the text "Lxy", where:
* **Lx** - number of difficulty Level
* **y** - "A" (answers) or "T" (tasks)

## Levels:
* L1 - very simply
* L2 - simple
* L3 - difficult
* L4 - very difficult
* L5 - very very difficult (real effective solution)

## Structure of the main notebooks:
* Import libraries
* Download data
* EDA & FE - Levels 1-3
* Modeling - Levels 1-7
* Prediction & Submission

## EDA & FE:
* Level 1:
        * view data, 
        * simple plot, 
        * describe data, 
        * replacement missing data for constanta,
        * creation a new features with simple functions, 
        * selection a some features
       
       
* Level 2:
        * the same and more plots with seaborn,
        * filter anomalous data, 
        * using EDA libraries: pandas-profiling, AutoViz
        * creation a new features with the way F1.astype('str') + F2.astype('str') 
          and encoding str to number
        * selection features by diagram feature importance
       
       
* Level 3:
        * the same and more plots with seaborn, plotly,
        * clustering data,
        * creation features with a deep understanding of the subject area,
        * EDA with yellowbricks library and others, 
        * Automatic FE with sklearn, Autofeat and others       

## Modeling:
* Level 1:
        * build a one model without parameters tuning (or using pre-training model) 
          without cross-validation (sklearn, fbprophet),
        * confusion matrix
        
        
* Level 2:
        * build a one model with parameters tuning (or using pre-training model)
          without cross-validation (sklearn, fbprophet, xgboost, lightgbm),
        * confusion matrix
        * different metrics
       
       
* Level 3:
        * build a one model with parameters tuning (or using pre-training model)
          and with cross-validation (sklearn, fbprophet, xgboost, lightgbm, keras, tensorflow, pytorch),
        * confusion matrix,
        * building learning curve of model
        
       
* Level 4:
        * the same and more - models ensemble
        * outcames analysis
       
       
* Level 5 - very difficult solutions for real tasks and Kaggle competitions:
        * tabular prediction
        * time series foracasting
        * image recognition or classification
        * NLP
        * etc.
