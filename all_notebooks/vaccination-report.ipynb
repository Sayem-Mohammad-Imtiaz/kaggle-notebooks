```python
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
```

```python
df=pd.read_csv('/kaggle/input/covid-world-vaccination-progress/country_vaccinations.csv')
df.head()
```

## What vaccines are used and in which countries

```python
countries = df['country'].unique()
d = {}
for i in df.values:
    d[i[0]] = d.get(i[0],[])
    if i[12] not in d[i[0]]:
        d[i[0]].append(i[12])
for i,j in d.items():
    print(i," : ",j)
```


## What country is vaccinated more people? 

```python
df_country = df.groupby(['country'])['people_vaccinated'].max()
df_country=df_country.sort_values(ascending=False)[:10]
plt.bar(df_country.index,df_country.values)
plt.xticks(rotation=90)
plt.show()
```

## What country is vaccinated a larger percent from its population?

```python
df_country = df.groupby(['country'])['people_fully_vaccinated_per_hundred'].max()
df_country=df_country.sort_values(ascending=False)[:10]
plt.bar(df_country.index,df_country.values)
plt.xticks(rotation=90)
plt.show()
```

## Daily vaccination per Million

```python
df_country = df.groupby(['country'])['daily_vaccinations_per_million'].max()
df_country=df_country.sort_values(ascending=False)[:10]
plt.bar(df_country.index,df_country.values)
plt.xticks(rotation=90)
plt.show()
```

## vaccination Report per day **World Wide**

```python
df_date = df.groupby('date')['total_vaccinations'].sum()
df_date = df_date.sort_values()[:10]
plt.plot(df_date.index,df_date.values)
plt.xticks(rotation=90)
plt.show()
```

## vaccination Report per day **Country**

```python
df_country = df.groupby(['country'])['total_vaccinations'].max()
df_country=df_country.sort_values(ascending=False)[:10]
plt.bar(df_country.index,df_country.values)
plt.xticks(rotation=90)
plt.show()
```
