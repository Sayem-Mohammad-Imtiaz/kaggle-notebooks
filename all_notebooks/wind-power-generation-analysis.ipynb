```python
wind_power_generation = pd.read_csv("../input/taiwan-power-company-wind-power-generation/_.csv")
wind_power_generation
```

```python
wind_power_generation.info()
```

```python
#將年度.縣市分組
county = wind_power_generation.groupby(['縣市/City'])
county.head(10)
```

```python
# 2021年各縣市的發電機資料筆數
year_2021 = wind_power_generation[wind_power_generation['年度/Year'] == 2021]  # 選出2021年的群組 
cities = year_2021['縣市/City'].unique()
print(f"2021年各縣市的發電機資料筆數：")
for city in cities:
    no =  len(year_2021.loc[year_2021['縣市/City'] == city])
    print(f"{city}     筆數：{no}")
```

```python
#2020年度各縣市的最多發電量度數
year_2020 = wind_power_generation[wind_power_generation['年度/Year'] == 2020]  # 選出2020年的群組 
cities = year_2020.groupby(['縣市/City'])['風機發電量(度)'].max()
print(f"2020年度各縣市的最多發電量度數: \n{cities}")
```

```python
#2016年台中市的發電機風機故障時數(小時)大於風機發電時數(小時)：
wind_power_generation.columns = ['年度/Year','月份/Month','縣市/City','風站名稱/Site Name','風機編號/Wind Turbine Number','裝置容量KW/Installed Capacity KW','風機發電量','風機發電時數','風機故障時數']
year_2016 = wind_power_generation[wind_power_generation['年度/Year'] == 2016]
year_2016['風機故障時數'] = pd.to_numeric(year_2016['風機故障時數'], errors='coerce')
year_2016['風機發電時數'] = pd.to_numeric(year_2016['風機發電時數'], errors='coerce')
taichang = year_2016['縣市/City'].str.contains("台中市")
taichang_break = year_2016[taichang]
taichang_break 
taichang_breakthanwork = taichang_break[taichang_break['風機故障時數'] >= taichang_break['風機發電時數']]
print(f"2016年台中市的發電機風機故障時數(小時)大於風機發電時數(小時)：")
display(taichang_breakthanwork)
```
