# Basic analysis of COVID-19 vaccination in Tamil Nadu Districts

```python
vacc = pd.read_csv('../input/covid-19-vaccination-coverage-across-tamilnadu/health_unit_district_wise_covid_19_vaccination_in_tamilnadu_as_on_30_05_2021.csv')
print(vacc.info())
```

```python
vacc.drop(labels=[45], axis=0, inplace = True)
```

## Covishield

```python
plt.figure(figsize=(20,20))
x = vacc['Health Unit District']
y = vacc['Achievement towards vaccination of 1st Dosage Covishield to HCW']
plt.subplot(2, 2, 1)
plt.bar(x,y)
plt.title('1st vaccination of HWC')
plt.xticks(rotation=90)
plt.ylim([0, 60])

x = vacc['Health Unit District']
y = vacc['Achievement towards vaccination of 2nd Dosage Covishield to HCW']
plt.subplot(2, 2, 2)
plt.bar(x,y)
plt.title('2nd vaccination of HWC')
plt.xticks(rotation=90)
plt.ylim([0, 60])
plt.show();
```
