Data Preparation

```python
dataset.info()
```

```python
dataset.drop(['ID'],axis = 1,inplace=True)
```

```python
dataset.head
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
dataset_std = scaler.fit_transform(dataset)
```

```python
dataset_std
```

Modelling

```python
#Hierarchical Clustering
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(dataset_std,method = 'ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 4, affinity = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(dataset_std)
```

```python
y_hc
```

```python
dataset['Cluster Name']=y_hc
```

```python
dataset.groupby(by = 'Cluster Name').mean()
```

```python
#K means clustering 
from sklearn.cluster import KMeans
```

```python
dummy = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(dataset)
    dummy.append(kmeans.inertia_)
plt.plot(range(1, 11), dummy)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
km = KMeans(n_clusters = 4, init = 'k-means++',random_state = 42)
y_kmeans = km.fit_predict(dataset_std)
```

```python
y_kmeans
```

```python
dataset['Cluster Name']=y_kmeans
dataset.groupby(by = 'Cluster Name').mean()
```
