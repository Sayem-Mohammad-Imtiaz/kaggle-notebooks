```python
import pandas as pd
import os
```

### Import the data

```python
all_data = pd.read_csv("../input/ligue-1-saison-20202021/data_2020_2021.csv")
```

### adding a new colum for all goals 

```python
all_data['All goals']= all_data['home_goal'] +all_data['away_goal']
all_data.head(10)
```

### finding the avrege the max goal scored per game and the total goal scored during the saison

```python
all_goals=all_data['All goals']

df=pd.DataFrame({'The avreg goal scored per game':[all_goals.mean()],
                'all goals scored in the league':[all_goals.sum()],
                'most goal scored in a game':all_goals.max()})
df
```

### finding avrege and most goal scored in home games and total home goal scored

```python
home_goals=all_data['home_goal']

dt=pd.DataFrame({'The avreg goal scored per game home':[home_goals.mean()],
                'all goals scored in the league from home staduime':[home_goals.sum()],
                'most goal scored in a home game ':home_goals.max()})
dt


```

### finding avrege and most goal scored in away games and total away goal scored

```python
away_goals=all_data['away_goal']

da=pd.DataFrame({'The avreg goal scored in game away':[away_goals.mean()],
                'all goals scored in the league from away staduime':[away_goals.sum()],
                'most goal scored in a away game ':away_goals.max()})
da

```

### finding wich journee hade the most goal (wich week)

```python
import matplotlib.pyplot as plt 

result = all_data.groupby('journee').sum()

```

```python
import matplotlib.pyplot as plt 

jour = range(1,39)

plt.plot(jour,result['All goals'])
plt.xticks(range(1,39),size=8)
from pylab import rcParams   #from stackover flow 
rcParams['figure.figsize'] = 16, 20


plt.xlabel('Journee',size=20)
plt.ylabel('Goals scored in that Journee',size=20)
plt.grid()
plt.show()
```

### HOME GOALS

```python
results_teams=all_data.groupby('Team').sum()
teams =[team for team,df in all_data.groupby('Team')]
plt.bar(teams,results_teams['home_goal'])
rcParams['figure.figsize'] = 8, 10
plt.xticks(teams,rotation='vertical',size=8)
plt.show()
```

### relation between home goals and the all goals

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(teams,results_teams['All goals'])
ax2.plot(teams,results_teams['home_goal'],'black')

ax1.set_xlabel('Teams')
ax1.set_ylabel('All Goals', color='b')
ax2.set_ylabel('Home Goals', color='black')
ax1.set_xticklabels(teams,rotation='vertical',size=8)

fig.show()
```
