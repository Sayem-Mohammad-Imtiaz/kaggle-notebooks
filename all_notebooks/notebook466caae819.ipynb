Nesta etapa foi importado o dataset para análise dos dados.

Foi removido das colunas PassengerId e Category.

Foi traduzido as colunas e valores da coluna Sobrevivente

A função head retorna as primeiras 5 linhas do objeto com base na posição.

```python
data=pd.read_csv('/kaggle/input/passenger-list-for-the-estonia-ferry-disaster/estonia-passenger-list.csv')
data=data.drop(columns=["PassengerId"])
data=data.drop(columns=["Category"])
data.columns = ['País', 'Nome', 'Sobrenome', 'Sexo', 'Idade', 'Sobrevivente']
data["Sobrevivente"] = data["Sobrevivente"].replace(to_replace=0, value= "Não")
data["Sobrevivente"] = data["Sobrevivente"].replace(to_replace=1, value= "Sim")
data.head()
```

A função describre gera estatísticas descritivas, que incluem aquelas que resumem a tendência central, a dispersão e a forma da distribuição de um conjunto de dados, excluindo os valores NaN (Not a number).

```python
print(data.País.describe())
```

```python
print(data.Idade.describe())
```

A função mean retorna a média dos valores para a coluna escolhida.

A função std retorna o desvio padrão para a coluna escolhida.

A função value_counts retorna uma série contendo contagens de linhas exclusivas.

A função groupby retorna o agrupamento dos dados se baseando em certos critérios.



```python
print("Média:", data.Idade.mean())
print("Desvio padrão:", data.Idade.std())
print("Contagem por sexo:\n",data.Sexo.value_counts())
data.groupby(['Sexo']).mean()
```

O gráfico de pizza abaixo mostra a proporção de pessoas que sobreviveram e não sobreviveram.

```python
data["Sobrevivente"].value_counts().plot.pie(title="Número de Sobreviventes")
```

O gráfico de barras abaixo mostra a distribuição de pessoas por país.

```python
data.País.value_counts().plot.bar()
```

O gráfico é do tipo histograma e mostra a distribuição de pessoas por idade.

```python
data.Idade.plot.hist()
```
