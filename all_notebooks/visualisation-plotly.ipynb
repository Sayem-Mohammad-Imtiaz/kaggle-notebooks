```python
import numpy as np 
import pandas as pd 
import collections

from plotly.offline import init_notebook_mode, iplot, plot

import plotly.express as px
init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.graph_objs as go
import missingno as msno
```

```python
books_data = pd.read_csv('/kaggle/input/amazon-top-50-bestselling-books-2009-2019/bestsellers with categories.csv')
books_data.head(5)
msno.bar(books_data)
```

No null values in any columns so missing data handling required.

```python
author_ratings_df = books_data.groupby('Year')['User Rating'].apply(list)
author_ratings_df.to_frame()
author_ratings_df = author_ratings_df.reset_index()
occurrences = []
for i in range(author_ratings_df.shape[0]):
    occurrences.append({'year':author_ratings_df['Year'].iloc[i],'count':dict(collections.Counter(author_ratings_df['User Rating'].iloc[i]))})
years_df = []
for ele in occurrences:
    year_ratings_df = pd.DataFrame.from_dict(ele)
    year_ratings_df.reset_index(level=0,inplace=True) 
    year_ratings_df.columns = ['rating','year','count']
    years_df.append(year_ratings_df)
    
year_ratings_df = pd.concat(years_df)
year_ratings_df.reset_index(inplace=True)
year_ratings_df['year'] = year_ratings_df['year'].map(str)
year_ratings_df['rating'] = year_ratings_df['rating'].map(str)
year_ratings_df = year_ratings_df.drop(['index'],axis=1)

fig = px.bar(year_ratings_df, x="year", y="count", color="rating", title="Ratings Distribution each Year")
fig.show()
```
