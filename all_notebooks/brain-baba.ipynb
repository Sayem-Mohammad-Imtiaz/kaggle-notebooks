```python
 %matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (20.0, 10.0)
data = pd.read_csv('/kaggle/input/headbrain/headbrain.csv')
print(data.shape)
```

```python
data.head(5)
```

```python
X=data['Head Size(cm^3)'].values
Y=data['Brain Weight(grams)'].values
```

```python
mean_x=np.mean(X)
mean_y=np.mean(Y)

n=len(X)

numer=0
denum=0

for i in range(n):
    numer += (Y[i]-mean_y)*(X[i]-mean_x)
    denum+= (X[i]-mean_x)**2
    
m=numer/denum
print(m)
```

```python
c=mean_y-m*mean_x
print(c)
```

```python
max_x = np.max(X) + 100 
min_x = np.min(X) - 100
x = np.linspace(min_x, max_x, 1000)
y = c + (m * x)
```

```python
plt.scatter(X, Y, color='#ef4423', label='Scatter points')
plt.plot(x, y, color='#52b920', label='Regression line')
plt.xlabel('Head size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()
```

```python
rs_num=0
rs_denum=0
for i in range(n):
    y_pred=m*X[i]+c
    rs_num+= (Y[i]-y_pred)**2
    rs_denum+=(Y[i]-mean_y)**2
r_squared=1-(rs_num/rs_denum)    
print(r_squared*100,"percent")
```
