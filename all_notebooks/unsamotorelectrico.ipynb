**Información**


Todos los registros se muestrean a 2 Hz (una cada 0,5 segundos). El dataset consta de varias sesiones de medición, que se pueden distinguir por la columna "profile_id". Una sesión de medición puede durar entre una y seis horas.

Las columnas "motor_speed" y "torque" son las cantidades  derivadas de las corrientes y voltajes establecidos.

El motor se enciende mediante ciclos diseñados a mano que indican una velocidad de motor de referencia y un torque de referencia.

Las corrientes en las coordenadas d/q (columnas "id" e iq") y los voltajes en las coordenadas d/q (columnas "ud" y "uq") son el resultado de una estrategia de control estándar que intenta seguir la velocidad y el torque .

La mayoría de los ciclos  denotan random walks para imitar los ciclos de conducción del mundo real 



**Objetivo**

Los objetivos más interesantes son la temperatura del rotor("pm"), las temperaturas del estator ("estator_*") y el "torque". Especialmente la temperatura del rotor y torque no se pueden medir de forma fiable y económica en un vehículo comercial.

Poder tener estimadores sólidos para la temperatura del rotor ayuda a la industria automotriz a fabricar motores con menos material y permite estrategias de control para utilizar el motor a su máxima capacidad.

Una estimación precisa del torque lleva a un control más preciso y adecuado del motor, lo que reduce las pérdidas de potencia y, finalmente, la acumulación de calor.
