```python
df=df.drop(['bike_id','start_station_id','start_station_latitude','start_station_longitude','end_station_id','end_station_latitude','end_station_longitude'],axis=1)
```

```python
df['start_time'] = pd.to_datetime(df['start_time'], errors='coerce')
df['end_time'] = pd.to_datetime(df['end_time'], errors='coerce')
```

```python
df['start_dayofweek'] = df.start_time.dt.strftime('%A')
df['start_month'] = df.start_time.dt.strftime('%B')
df['end_dayofweek'] = df.end_time.dt.strftime('%A')
df['end_month'] = df.end_time.dt.strftime('%B')
```

```python
df.head()
```

```python
fig_dims = (10,6)
fig,ax = plt.subplots(figsize=fig_dims)
sns.countplot(x='start_month',data = df,ax=ax)
```

```python
fig_dims = (10,6)
fig,ax = plt.subplots(figsize=fig_dims)
sns.countplot(x='end_month',data = df,ax=ax)
```

```python
pie_df = df.user_type.value_counts().reset_index()
pie_df.columns = ['user_type','count']
fig = px.pie(pie_df, values='count', names='user_type', title='Proportion of user_type',
             color_discrete_sequence=['#2A4B7C','#E08119','#E9897E'])
fig.show()
```
