```python
import sklearn
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
train_df=pd.read_csv('/kaggle/input/diabetes-data-set/diabetes-dataset.csv')
```

```python
train_df.info()
```

```python
train_df_features=train_df.iloc[:,:8]
train_df_targets=train_df.iloc[:,8]
```

```python
train_df,test_df,train_tar,test_tar=model_selection.train_test_split(train_df_features,train_df_targets,test_size=0.2,shuffle=True)
```

```python
train_df.head()
```

```python
train_df.reset_index(drop=True,inplace=True)
test_df.reset_index(drop=True,inplace=True)
```

```python
print(f"The Shape of the train dataset is {train_df.shape}")
print(f"The shape of the test dataset is {test_df.shape}")
print(f"The Shape of the train dataset is {train_tar.shape}")
print(f"The shape of the test dataset is {test_tar.shape}")
```

Converting the pandas dataframe to numpy

```python
train_df=np.array(train_df)
test_df=np.array(test_df)
train_tar=np.array(train_tar)
test_tar=np.array(test_tar)
train_tar=train_tar.reshape(train_tar.shape[0],-1)
test_tar=test_tar.reshape(test_tar.shape[0],-1)
```

```python
sc=StandardScaler()
```

```python
train_df=sc.fit_transform(train_df)
```

**Fitting Classifiers on this Data**

# **SVM*****

```python
clf=SVC(gamma='auto',kernel='')
```

```python
train_tar=train_tar.ravel()
```

```python
clf.fit(train_df,train_tar)
```

```python
preds=clf.predict(test_df)
print(preds.shape)
```

```python
from sklearn.metrics import accuracy_score
```

```python
test_tar=test_tar.ravel()
```

```python
accuracy=accuracy_score(test_tar,preds)
print(accuracy)
```
