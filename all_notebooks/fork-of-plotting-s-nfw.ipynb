```python
# Defining constant 'e' and solar mass Ms

e = float(84.862746)
Ms = float(1.989 * (10**(10)))

# convert H0 from km/sec/Mpc to /sec
H0 = 70*(3.24e-20)
```

```python
def snfw(m, c, z):
    cuberoot_value = (float(m * Ms))**(1/3) 
    arctan_value = np.arctan (math.sqrt(c*c-1))
    log_value = math.log((c/2),10)
    Hz = H0 * math.sqrt(0.7 + (0.3 * ((1+z)**3)))
    rhocrit = 3*Hz*Hz/(8*np.pi/G)
    A = np.log(1+c)-c/(1+c)
    s_nfw = e * (1/A) * cuberoot_value * (rhocrit**(2/3)) * (log_value + (arctan_value * ((c*c-1)**(-1/2)))) 
    return(s_nfw)
```
