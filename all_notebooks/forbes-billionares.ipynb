```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
path = "../input/forbes-billionaires-of-2021-20/forbes_billionaires.csv"
data = pd.read_csv(path)
data.head()
```

```python
#rTop 20 Richest People In the World location  'RPL'
rpl = data.value_counts('Country')
rpl[:20]
```

```python
rpl_index =rpl.index
rpl_index[:20]
```

```python
plt.figure(figsize=[16,6])
order = rpl_index[:20]
sns.countplot(data = data , x = 'Country' ,
             order = order)
plt.xticks(rotation = 90)
plt.title('Top 20 Contries by number of rich People ', fontname='Time New Romans' , fontsize = '30')
plt.ylabel('number of rich People', fontname='Time New Romans' , fontsize = '20')
plt.show()
```

```python
# top 20 contries by totale of Billions of  $ 'cpt'
cpt = data.groupby('Country')['NetWorth'].sum().sort_values(ascending=False)

cpt_data=pd.DataFrame(cpt)
cpt_t10=cpt_data[:10]
cpt_t10
```

```python
plt.figure(figsize=[16,6])

plt.bar(x=cpt_t10.index, height=cpt_t10['NetWorth'])

plt.xticks(rotation = 90)
plt.title('Top 10 Contries by Total of Billions of $ ', fontname='Time New Romans' , fontsize = '30')
plt.ylabel('Total of Billions of $ ', fontname='Time New Romans' , fontsize = '20')
plt.show()
```

```python
# A view on how many billionairs have more than 10 B 
more_than_10 = data[data.NetWorth>10].value_counts().sum()
percent = more_than_10/len(data)*100
print(len(data) )
print( more_than_10)
print('percent of billionaire with more than 10Billions ',"{:.2f}".format(percent),'%')
```

```python
# and as always most of them still in Usa and China
data[data.NetWorth>10]['Country'].value_counts()[:10]
```

```python
plt.figure(figsize=[16,6])
data_top_contries = data[data['Country'].isin(['United States','China','Germany','India','Russia'])]

sns.swarmplot(data=data_top_contries, x = 'Country', y = 'NetWorth', hue = 'Self_made',dodge=True)

plt.show()
```
