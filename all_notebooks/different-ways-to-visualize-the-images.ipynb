```python
pd.set_option('display.max_colwidth', -1)

def get_thumbnail(path):
    i = Image.open(path)
    i.thumbnail(size=(150, 150), resample=Image.LANCZOS)
#     i.thumbnail(size=(150, 150))
    return i

def image_base64(im):
    if isinstance(im, str):
#         print('Inside isinstance')
        im = get_thumbnail(im)
#         print('got im')
    with BytesIO() as buffer:
#         print('inside with block, and saving the file')
        im.save(buffer, 'jpeg')
#         print(base64.b64encode(buffer.getvalue()).decode())
        return base64.b64encode(buffer.getvalue()).decode()

def image_formatter(im):
    return f'<img src="data:image/jpeg;base64,{image_base64(im)}">'
```

# 3. Images from Folder - Random x Images

```python
# show some random images
plt.figure(figsize=(12,12))

for k in range(28):
    randomImg = random.choice(
    [
        x for x in os.listdir(train_image_dir)
        if os.path.isfile(os.path.join(train_image_dir, x))
    ]
)
    
    imageFilename = train_image_dir + str(randomImg) 
    
    plt.subplot(4,7,k+1); 
    
    plt.imshow(mpimg.imread(imageFilename), cmap='gray')

    plt.axis('off')
```
