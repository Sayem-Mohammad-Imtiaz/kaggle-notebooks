```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score


train_df = pd.read_csv('../input/train.csv')
train_df.head(500)
```

```python
X_train, y_train = train_df.iloc[:, 0:len(train_df.columns) - 1], train_df.iloc[:, -1]
```

```python
test_df = pd.read_csv('../input/test.csv')
test_df.head()
```

```python
X_test, y_test = test_df.iloc[:, 0:len(test_df.columns) -1], test_df.iloc[:, -1]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 2019)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
```

```python
X_test['Id'] = range(1, len(X_test) + 1)
X_test.head()
```

```python
sub = pd.DataFrame()
sub['Id'] = X_test['Id']
sub['Predicted Value'] = y_pred
sub['Actual value'] = y_test
sub.head()
```

```python
print('Accuracy score:', accuracy_score(y_test, y_pred))
```

```python
sub.to_csv('submission.csv',index=False)
```
