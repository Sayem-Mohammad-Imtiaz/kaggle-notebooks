# Data Analysis & Visualization
Here we have done the following things:

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from scipy import stats

import warnings
warnings.filterwarnings('ignore')
```

Here we have created two functions to make our visualization work easy

```python
# As there are total of 47 countries, we split it into two sections in a function
def first_27(df, catagory): 
    df = df.sort_values(catagory, ascending = False)
    fig = px.bar(df.head(27), y = catagory, x = 'Country',
            text = catagory, color = 'Country')
    return fig

def last_20(df, catagory): 
    df = df.sort_values(catagory, ascending = False)
    fig = px.bar(df.tail(20), y = catagory, x = 'Country',
            text = catagory, color = 'Country')
    return fig
```

## Total Cases per Country
`NOTE:` As there are total 47 countries, we have splited the graph into two parts. First one is for the first 27 countries and the second one is for the last 20 countries.

```python
first_27(df, 'Total Cases')
```

```python
last_20(df, 'Total Cases')
```
