**Lets Scrutinize the Female Gender now **

```python
female_count = len(df_mcq.query("GenderSelect=='Female'"))
male_count = len(df_mcq.query("GenderSelect=='Male'"))
```

```python
for idx,count in zip(df_mcq.query("GenderSelect=='Female'")["EmploymentStatus"].value_counts().index,df_mcq.query("GenderSelect=='Female'")["EmploymentStatus"].value_counts()) : 
    print(idx,":",format((count*100)/female_count,".2f"))
```

**Observations ** 

- 59% of the Females are employed full time . 
- Around 17% of them are on a job-hunt 
- Around 7% of the females are pretty happily settled , they don't really care about jobs . 
- 6% of the females work for themselves 
- 3.25 of them have chosen to not speak of their profession . (But why ?)
- And finally a few have retired . 

**Male Gender basic scrutiny **

```python
for idx,count in zip(df_mcq.query("GenderSelect=='Male'")["EmploymentStatus"].value_counts().index,df_mcq.query("GenderSelect=='Male'")["EmploymentStatus"].value_counts()) : 
    print(idx,":",format((count*100)/male_count,".2f"))
```

**Observations ** 

- Around 67% of the males are employed full time . 
- Around 12% of them are on a job-hunt 
- Around 8.5% of the males work for themselves . 
- 5.25% % of them seem well off .  
- 2.16% of them have chosen to not speak of their profession . (hmmm ?)
- And finally a few have retired . 

**Lets drill down the facts location-wise **

```python
df_mcq['Country'].value_counts(20)
```

**Lets , for the time being just consider the top 20 countries **

```python
df_mcq_top20 = df_mcq[df_mcq['Country'].isin(df_mcq['Country'].value_counts().head(20).index)]
```

```python
plt.figure(figsize=(20,15))

g = sns.countplot(df_mcq_top20['Country'])
locs, labels = plt.xticks()
g.set_xticklabels(labels,rotation=45)
```

```python
plt.figure(figsize=(20,15))

g = sns.boxplot(x=df_mcq_top20["Country"],y=df_mcq_top20["Age"])

plt.legend(loc="upper right")
plt.ylabel("Age")
plt.xlabel("Countries")
locs, labels = plt.xticks()
g.set_xticklabels(labels,rotation=45)
plt.tight_layout()
```

**The age band seems pretty uniform except for India, China and Ukraine where in there are a few younger people who have been involved as a part of the analysis**

```python
plt.figure(figsize=(12,10))
g = sns.boxplot(x=df_mcq['CurrentJobTitleSelect'],y=df_mcq["Age"],hue=df_mcq['TitleFit'])
plt.legend(loc="upper right")
locs, labels = plt.xticks()
g.set_xticklabels(labels,rotation=90)
plt.tight_layout()
```
