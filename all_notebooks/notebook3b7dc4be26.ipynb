```python
# Train_Test Split

feature_col = ['age','anaemia','creatinine_phosphokinase','diabetes','ejection_fraction','high_blood_pressure','platelets','serum_creatinine','serum_sodium','sex','smoking']
X = df[feature_col].values
y = df['DEATH_EVENT']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size = 0.3,random_state = 58)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#Model
rf = RandomForestClassifier(n_estimators = 100,criterion = 'gini',oob_score = True,random_state = 75
                            )
rf.fit(X_train,y_train)
```

```python
y = preprocessing.label_binarize(y, classes=[0, 0, 0, 0])
predicted = rf.predict(X_test)
rfc_cv_score = cross_val_score(rf, X, y, cv=10)
print("=== Confusion Matrix ===")
print(confusion_matrix(y_test, predicted))
print('\n')
print("=== Classification Report ===")
print(classification_report(y_test, predicted))
print('\n')
print("=== All AUC Scores ===")
print(rfc_cv_score)
print('\n')
print("=== Mean AUC Score ===")
print("Mean AUC Score - Random Forest: ", rfc_cv_score.mean())
```
