```python
raw_data = pd.read_csv('/kaggle/input/insurance/insurance.csv' , true_values=['yes' , 'female'] , false_values=['no' , 'male'] , dtype={'somker':int})
raw_data["sex"] = raw_data["sex"].astype(int)
raw_data["smoker"] = raw_data["smoker"].astype(int)
raw_data
```

```python
processed_data = pd.get_dummies(raw_data , columns = ['region'] , drop_first=True) #converting the region feature to 3 binary features
processed_data
```

```python
y = processed_data['charges']
x = processed_data.loc[:, processed_data.columns != 'charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15)
```

```python
from sklearn.preprocessing import MaxAbsScaler
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from sklearn.decomposition import PCA
```

```python
steps = [('scaler', MaxAbsScaler()),
         ('poly feature', PolynomialFeatures(degree=3)),
         #('dimension reduction', PCA(n_components = 0.96)),
         #('Regression', LinearRegression()),
         ('Regression' , Ridge(0.9))
         ]
my_pipeline = Pipeline(steps=steps)
my_pipeline.fit(x_train, y_train)
print('test score: {}'.format(my_pipeline.score(x_test, y_test)))
print('train score: {}'.format(my_pipeline.score(x_train, y_train)))
```
