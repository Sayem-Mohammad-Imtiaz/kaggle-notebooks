```python
import plotly.graph_objects as go
import numpy as np


layout = go.Layout(title_text='Correlation Matrix',
                   title_x=0.5, 
                   width=600, height=600,
                   xaxis_showgrid=False,
                   yaxis_showgrid=False,
                   yaxis_autorange='reversed')

# Create figure
fig = go.Figure(layout=layout)


# Add traces, one for each slider step
for step, df in dfs.items():
    corr = df.corr()

    labels = df.columns
    hovertext = [
        [(f'corr({i_val}, {j_val})= {corr.iloc[i,j]:.2f}' if i>j and i!=j else '')\
            for j, j_val in enumerate(corr.columns)] \
            for i, i_val in enumerate(corr.columns)
    ]

    heat = go.Heatmap(z=corr.values,
                      x=labels,
                      y=labels,
                      xgap=1, ygap=1,
                      colorbar_thickness=20,
                      colorbar_ticklen=3,
                      hovertext =hovertext,
                      hoverinfo='text')
    trace = heat
    fig.add_trace(trace)


# Make 10th trace visible
fig.data[0].visible = True

# Create and add slider
steps = []
for i in range(len(fig.data)):
    step = dict(
        method="restyle",
        args=["visible", [False] * len(fig.data)],
    )
    step["args"][1][i] = True  # Toggle i'th trace to "visible"
    steps.append(step)

sliders = [dict(
    active=10,
    currentvalue={"prefix": "Year: "},
    pad={"t": 50},
    steps=steps
)]

fig.update_layout(
    sliders=sliders
)

fig.show()
```
