```python
base_path = '/kaggle'
feature_dir = f"{base_path}/input/unified-ds-wifi-and-beacon"
train_files = sorted(glob.glob(os.path.join(feature_dir, '*_train.csv')))
test_files = sorted(glob.glob(os.path.join(feature_dir, 'test.csv')))
```

```python
RSSI_FEATS  = [f'wifi_rssi_{i}' for i in range(100)]
BEACON_TIMEGAP = [f'beacon_timegap_{i}' for i in range(10)]
WIFI_TIMEGAP = [f'wifi_timegap_{i}' for i in range(100)]
BEACON_DISTANCE = [f'beacon_distance_{i}' for i in range(10)]
```

```python
data = None
for n_files, file in enumerate(train_files):
    file = file.replace('\\', '/')
    print(file)
    
    if data is None:
        data = pd.read_csv(file)
        print(data.shape)
    else:
#         if 'Unnamed: 0' in data.columns:
#             data.drop(['Unnamed: 0'],axis=1,inplace=True)
#         wifi_file = np.loadtxt(file, delimiter=',', dtype='str')
        wifi_file = pd.read_csv(file)
#         print(wifi_file.shape)
#         wifi_file = pd.DataFrame(wifi_file)
        wifi_file = wifi_file.iloc[1:,:]
        wifi_file.columns = ['Unnamed: 0', 'site', 'floor', 'path', 'x', 'y', ] + [f'wifi_bssid_{i}' for i in range(100)] + [f'wifi_rssi_{i}' for i in range(100)] + [f'wifi_timegap_{i}' for i in range(100)] +[f'beacon_macaddress_{i}' for i in range(10)] + [f'beacon_distance_{i}' for i in range(10)] + [f'beacon_timegap_{i}' for i in range(10)]
#         for col in RSSI_FEATS + ['floor']:
#         wifi_file[col] = wifi_file[col].astype('int')
            
#         for col in BEACON_DISTANCE + BEACON_TIMEGAP + WIFI_TIMEGAP:
#             wifi_file[col] = wifi_file[col].astype('float')
            
        wifi_file['x'] = wifi_file['x'].astype('float')
        wifi_file['y'] = wifi_file['y'].astype('float')
 
        data = pd.concat([data, wifi_file])
        data.reset_index(drop=True, inplace=True)
        print(data.shape)
```
