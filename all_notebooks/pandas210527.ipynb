# 9.Data Framelerin Birleştirmesi

```python
s1 = np.random.randint(10, size = 5)   
s2 = np.random.randint(10, size = 5)
s3 = np.random.randint(10, size = 5)
df1 = pd.DataFrame({"col1": s1, "col2": s2,"col3": s3})  
df1
```

```python
df2 = df1 + 99  
df2
```

```python
pd.concat([df1,df2], axis = 0)    # olusturulan iki adet df i satir bazinda birlestirdi (altalta), indexler ile beraber
```

▪ axis=1, yatay/horizontal (sutun bazli) birlestirme

```python
pd.concat([df1,df2], axis = 1)  # olusturulan iki adet df i sutun bazinda birlestirdi,(yanyana) indexler ile beraber
```

 # ignore_index = True ile indexleri tek bir index yaptik.

```python
pd.concat([df1,df2], ignore_index = True)
```

```python
pd.concat([df1,df2], ignore_index = False)
```

```python
df2.columns = ["col1", "col2", "col4"]
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2], join = 'inner')  # birlestirmede farkli sutunlar da varsa o sutunlari almadan, iki df de ortak olan sutunlari alacak
```

pd.merge()

One-to-one / many-to-one / many-to-many

```python
df1 = pd.DataFrame({'calisanlar': ['Ali', 'Veli', 'Ayse', 'Fatma'],
                    'grup': ['Muhasebe', 'Muhendislik', 'Muhendislik', 'İK']})
df1
```

```python
df2 = pd.DataFrame({'calisanlar': ['Ayse', 'Ali', 'Veli', 'Fatma'],   #sutunlar df1 ile ayni 
                    'ilk_giris': [2010, 2009, 2014, 2019]})
df2
```

### one-to-one

```python
pd.merge()
```

```python
# df1 ile df2'yi birebir eşleştiriyoruz. 
# Yani; kişi isimleri referans olacak şekilde birleştirilecek
pd.merge(df1,df2)
```

'merge' fonksiyonu birleştirme işleminin hangi değişkene göre yapılacağını biliyor. Ama istersek biz bunu "on = 'calisanlar'" şeklinde yönlendirebiliriz.

```python
pd.merge(df1,df2, on = 'calisanlar')
```

### many-to-one

```python
df3 = pd.merge(df1,df2)  # df1 ile df2 yi birleştirerek df3'ü oluşturuyoruz
df3
```

```python
df4 = pd.DataFrame({'grup': ['Muhasebe', 'Muhendislik', 'İK'],
                   'mudur': ['Caner', 'Mustafa', 'Berkcan']})
df4
```

```python
pd.merge(df3,df4)
# Tekrarlı değerler olmasını da dikkate alarak birleştirme yaptı. 
# Bundan dolayı many-to-one diye isimlendirildi.
```

### many-to-many birleştirme

```python
df1
```

```python
df5 = pd.DataFrame({'grup': ['Muhasebe', 'Muhasebe', 'Muhendislik', 'Muhendislik', 'İK', 'İK'],
                   'yetenekler': ['matematik', 'excel', 'kodlama', 'linux', 'excel', 'yonetim']})
df5
```

```python
pd.merge(df1,df5)
# Muhasebeye karşılık çalışan ismi ali ve yetenekler matematik ve exel olduğundan,
# many-to-many birleştirme işlemi gerçekleşti.
```

```python
# Aynı şeyi ifade eden ama farklı değişken isimlerine sahip verilerimiz olsun. Yani ortak değişkenimiz olmasin.
df3 = pd.DataFrame({'name': ['Ayse', 'Ali', 'Veli', 'Fatma'],   #sutunlar df1 ile ayni 
                    'maas': [7000, 8000, 4500, 6500]})
df3
```

```python
df1
```

```python
# pd.merge(df1,df3)
# bu kod hata verecek, cunku birleştirme için referans alacağı ortak bir değişken yok
```

```python
# Yukarda verilen 'MergeError' kısmında çözüm önerileri sunuyor.
pd.merge(df1,df3, left_on = 'calisanlar', right_on = 'name')
```

```python
# isimler iki defa tabloda yer alıyor. Bu sorunu çözmek için
pd.merge(df1,df3, left_on = 'calisanlar', right_on = 'name').drop('name',axis = 1)
```
