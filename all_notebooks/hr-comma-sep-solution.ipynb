```python
new_df = df[["satisfaction_level", "average_montly_hours", "promotion_last_5years", "salary"]]
new_df.head()
```

```python
dummies = pd.get_dummies(new_df.salary)
dummies.head()
```

```python
merged = pd.concat([new_df, dummies], axis=1)
merged.head()
```

```python
final = merged.drop(["salary"], axis=1)
final.head()
```

```python
X = final
X.head()
```

```python
y = df.left
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(solver='lbfgs', max_iter=150)
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```
