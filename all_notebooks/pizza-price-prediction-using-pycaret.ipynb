```python
import pandas as pd
pizza=pd.read_csv('../input/pizza-price-prediction/pizza_v1.csv')
pizza.head()
```

```python
pizza.info()
```

```python
pizza['price_rupiah'] = pizza['price_rupiah'].str[2:]
pizza.head()
```

```python
import re 
pizza['price_rupiah'] = pizza['price_rupiah'].map(lambda x: re.sub(r'\W+', '', x))
pizza.head()
```

```python
pizza.info()
```

```python
pizza['price_rupiah'] = pizza['price_rupiah'].astype(int)
pizza.info()
```

```python
data = pizza.sample(frac=0.8, random_state=42)
data_unseen = pizza.drop(data.index)
data.reset_index(inplace=True, drop=True)
data_unseen.reset_index(inplace=True, drop=True)
print('Data for Modeling: ' + str(data.shape))
```

```python
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.regression import *
promo= setup(data =data, session_id=42,
             target = 'price_rupiah',
                  normalize = True, 
                  transformation = True, 
                  log_experiment = True,
                  remove_multicollinearity = True, #rop one of the two features that are highly correlated with each other
            categorical_features=[ 'size','extra_sauce', 'extra_cheese'],
            high_cardinality_features =['topping','variant','company'], 
             transform_target = True,
            train_size = 0.8)
```
