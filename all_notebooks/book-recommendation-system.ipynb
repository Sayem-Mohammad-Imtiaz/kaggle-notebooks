```python
books=pd.read_csv('/kaggle/input/7k-books-with-metadata/books.csv')
```

```python
books.head(5)
```

```python
books= books.drop(columns=['isbn10','subtitle','thumbnail','published_year','num_pages'])
```

```python
books.head(2)
```

```python
books.shape
```

```python
books.drop_duplicates(subset ="title",keep = False, inplace = True)
```

```python
books.shape
```

```python
books.dropna(inplace=True)
```

```python
books.shape
```

```python
books['description'] = books['description'].apply(lambda x:x.split())
```

```python
books.head(3)
```

```python
books['categories'] = books['categories'].apply(lambda x:x.split())
```

```python
books.head()
```

```python
books['tags'] = books['categories']+books['description']
```

```python
books.head(2)
```

```python
books['ratings_count'].value_counts()
```

```python
books_new=books[['isbn13','title','average_rating','tags']]
```

```python
books_new.head()
```

```python
books_new = books_new[books_new['average_rating']>=3.5]
```

```python
books_new = books_new[['isbn13','title','tags']]
```

```python
books_new.shape
```

```python
books_new['title'].unique()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
 books_new['tags'] = books_new['tags'].apply(lambda x: " ".join(x))
```

```python
vector = cv.fit_transform(books_new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
books_new.head()
```

```python
books_new.reset_index(inplace=True)
```

```python
books_new.head()
```

```python
new = books_new[['title','tags']]
```

```python
new.head()
```

```python
new.shape
```

```python
book_title=new['title']
```

```python
book_title
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    #print(distances)
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Harry Potter')
```
