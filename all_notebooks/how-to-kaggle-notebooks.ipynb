# How to Kaggle Notebooks

In diesem Notebook werden kurz die wichtigsten Grundlagen zur Verwendung von Kaggle Notebooks vorgestellt. Kaggle Notebooks ist eine Notebook-Umgebung in der Python-Code und Text in einem interaktiven Dokument geschrieben und entwickelt werden kann.

Kaggle Notebooks basiert auf [IPython/Jupyter-Notebooks](https://jupyter.org/), und wird sehr ähnlich zu diesen verwendet.

## Zellen-Grundlagen

Der Grundbaustein eines Notebooks sind Zellen, von denen es zwei Typen gibt:

*   Textzellen: Dieser Text befindet sich in einer davon. Mit einem Doppelklick oder `Enter` kannst du den Text hier editieren, am oberen Rand des Felds erscheinen dann auch verschiedene Formatierungstools. Überschriften fangen mit dem Zeichen # an, wie oben zu sehen. Der Editier-Modus kann mit `Shift + Enter` verlassen werden.

*   Codezellen: In diese kann ausführbarer Pythoncode geschrieben werden. Sie können mit einen Klick auf den Pfeil auf der linken Seite ausgeführt werden (erscheint wenn die Zelle aktiv ist), oder mit `Shift + Enter`. Die Ausgabe des letzten Ausdrucks wird nach Ausführung unter der Zelle angezeigt.

```python
1+5
```

Neue Zellen können einfach mit den beiden Buttons **+ Code** und **+ Markdown** unter jeder Zelle hinzugefügt werden. Alternativ kann mit den Shortcuts `b` oder `a` eine Codezelle unter (b=below) oder über der aktuellen (a=above) hinzugefügt werden. Die Shortcuts funktionieren nur, wenn die Zelle aktuell nicht im Editiermodus bearbeitet wird. Hier eine Auswahl an nützlichen Shortcuts:

* `b` Codezelle unterhalb der aktuellen Zelle einfügen (b=below)
* `a` Codezelle oberhalb der aktuellen Zelle einfügen (a=above)
* `m` Codezelle in Markdown umwandeln (m=markdown)
* `y` Markdownzelle in Code umwandeln (y=?)
* `dd` Zelle löschen (d=delete)
* `z` zuletzt gelöschte Zelle wieder herstellen
* `Enter` Zelle editieren
* `Esc` Editiermodus verlassen
* `Shift`+`Enter` Code-Zelle ausführen/Markdown-Zelle verlassen und zur nächsten Zelle

Weitere Shortcuts und Konfigurationsmöglichkeiten können mit `Strg`+`Shift`+`p` angezeigt werden.

Zellen können auch mit der kleinen Toolbar rechts oben an der Zelle mit den Pfeilen nach oben oder unten verschoben werden.

Werden Codezellen einmal ausgeführt, und das Notebook gespeichert, bleibt die angezeigte Ausgabe erhalten und kann später betrachtet werden:

```python
import matplotlib.pyplot as plt
plt.plot([1,5,10,20])
```

## Kaggle Runtime

Ein wichtiger Unterschied zu normalen IPython-Notebooks ist die von Kaggle bereitgestellte Runtime aus der Cloud. Zur Verwendung ist keine lokale Installation oder Konfiguration nötig, Kosten fallen ebenfalls nicht an.

Dies dauert im Normalfall wenige Sekunden. Das **Run** Menü oben im Fenster bietet einige Optionen zur Ausführungen, wichtig sind hier besonders **Run before** (führt alle Zellen vor der aktuellen aus) und **Restart & clear cell outputs**/**Factory rest** (wichtig, falls Schlimme Dinge (tm) passiert sind).

Auf der rechten Seite gibt es eine Sidebar mit Einstellungen zur Runtime und den eingebundenen Datensätzen. Die Sidebar kann mit der Schaltfläche oben rechts ein- und ausgeblendet werden.



Um vom Notebook auf das Internet zugreifen zu können (z.B. um Daten zu laden) muss der Internet-Schalter aktiviert sein. Für den Zugriff auf eine GPU oder TPU muss der entsprechende Accelerator aktiviert werden (die Runtime startet dann neu).
