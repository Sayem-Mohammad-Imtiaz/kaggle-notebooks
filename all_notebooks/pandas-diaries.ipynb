# 3.Different Ways Of Creating Dataframe

```python
#1.CREATING DATAFRAME USING DICTIONARY.....

weather_data = {
    'Day': ['1/1/2017','1/2/2017','1/3/2017'],
    'Temperature': [32,35,28],
    'Windspeed': [6,7,2],
    'Event': ['Rain', 'Sunny', 'Snow']
}
df = pd.DataFrame(weather_data)
df
```

```python
#2.CREATING DATAFRAME USING LIST .....

weather_data = [
    ('1/1/2017',32,6,'Rain'),
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',28,2,'Snow')
]
df = pd.DataFrame(data=weather_data, columns=['Day','Temperature','Windspeed','Event'])
df
```

```python
#3.CREATING DATAFRAME USING DICTIONARY IN SIDE THE LIST .....

weather_data = [
    {'Day': '1/1/2017', 'Temperature': 32, 'Windspeed': 6, 'Event': 'Rain'},
    {'Day': '1/2/2017', 'Temperature': 35, 'Windspeed': 7, 'Event': 'Sunny'},
    {'Day': '1/3/2017', 'Temperature': 28, 'Windspeed': 2, 'Event': 'Snow'},
    
]
df = pd.DataFrame(data=weather_data, columns=['Day','Temperature','Windspeed','Event'])
df
```

# 5.Handling missing data fillna dropna interpolate

```python
df = pd.read_csv('../input/w-e-a-t-h-e-r-dataset/Nyc_weather.csv')
df
```

```python
%matplotlib inline
df['Temperature'].plot.bar()
```

```python
new_df = df.fillna(0)
new_df
```

```python
df.isnull().sum()
```

```python
new_df = df.fillna({
        'WindSpeedMPH': 0,
        'Events': 'No Event'
    })
new_df
```

```python
new_df = df.fillna(method="ffill")
new_df
```

```python
new_df = df.dropna(thresh=2)
new_df
```
