## Races data

```python
races_df = pd.read_csv("/kaggle/input/formula-1-world-championship-1950-2020/races.csv")
```

```python
races_df
```

```python
print("Number of unique races:",len(races_df.loc[:,"name"].unique()))
```

```python
races = races_df.loc[:,"name"].unique()

hist_races = pd.DataFrame()
for race in races:
    hist_races = hist_races.append([[race,len(races_df[races_df.loc[:,"name"]==race])]])
hist_races.columns = ["race", "number"]
hist_races = hist_races.sort_values(by=["number"], ascending = False)
hist_races = hist_races.reset_index(drop = True)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = hist_races.race
students = hist_races.number.values
ax.bar(langs,students)
plt.xticks(rotation=90)
plt.show()
```

## Drivers

```python
drivers_df = pd.read_csv("/kaggle/input/formula-1-world-championship-1950-2020/drivers.csv")
```

```python
drivers_df
```

```python
drivers = drivers_df.loc[:,"nationality"].unique()

hist_nationality = pd.DataFrame()
for nationnality in drivers:
    hist_nationality = hist_nationality.append([[nationnality,len(drivers_df[drivers_df.loc[:,"nationality"]==nationnality])]])
hist_nationality.columns = ["nationality", "number"]
hist_nationality = hist_nationality.sort_values(by=["number"], ascending = False)
hist_nationality = hist_nationality.reset_index(drop = True)
```

Constructors

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = hist_nationality.nationality
students = hist_nationality.number.values
ax.bar(langs,students)
plt.xticks(rotation=90)
plt.show()
```

## Lap time

```python
lap_df = pd.read_csv("/kaggle/input/formula-1-world-championship-1950-2020/lap_times.csv")
#Italian grand prix - raceId 14
```

```python
monza_races_df = races_df[races_df.name == "Italian Grand Prix"]
monza_races_df = monza_races_df.sort_values(by = "year")
monza_races_df
```

```python
monza_df = pd.DataFrame()
monza_df["year"] = ""
for i in monza_races_df.index.values:
    monza_df = monza_df.append([lap_df[lap_df.raceId == monza_races_df.loc[i, "raceId"]]])
    #monza_df[monza_df.raceId == monza_races_df.loc[i, "raceId"]].loc["year"] = monza_races_df.loc[i, "year"]
```

```python
monza_df = monza_df.sort_values(by="milliseconds", ascending = False)
monza_df = monza_df.reset_index(drop = True)

fig = plt.figure()
ax = plt.axes()

x = monza_df.index.values[1000:]
ax.plot(x, monza_df.milliseconds[1000:]);

monza_df
```
