### 3. Movement restriction

3 Papers write about movement restriction.<sup>[6-8]</sup>

There have been reports that the number of people infected has been reduced by 50%<sup>[6]</sup> or 90%<sup>[7]</sup> due to movement restrictions, and that there is a strong positive correlation between the number of people affected and the number of people infected<sup>[8]</sup>.

### 4. Other risk factor

One article describes various risks of severe disease<sup>[9]</sup>. 

Risk factor described in paper is below.

 - history of smoking
 - maximum body temperature at admission
 - respiratory failure
 - Higher respiratory rate
 - Higher C-reactive protein
 - Lower Albumin

---

Other papers describe cardiovascular risk and severity<sup>[10]</sup>. 

Risk factor described in paper is below.

 - Higher C-reactive protein
 - Higher creatinine
 - Higher N-terminal pro B-type natriuretic peptide(NT-proBNP; only men)
 - Higher cardiac troponin-I (cTnl; only men)

---

Other papers describe age and underlying disease <sup>[11]</sup>.

Risk factor described in paper is below.

 - Higher age
 - Higher number of chronic underlying diseases

CRP is an indicator of the inflammatory response and may indicate the risk of severity.

In addition, it is highly likely that a healthy body is less likely to be severely ill.
