```python
df['Year_Month']=df['Date'].array.strftime('%Y-%m')
df.head()
```

```python
filter1=[
 'MinTemp',
 'MaxTemp',
 'Rainfall',
 'WindGustSpeed',
 'WindSpeed9am',
 'WindSpeed3pm',
 'Humidity9am',
 'Humidity3pm',
 'Pressure3pm',
 'Pressure9am',
 'Temp9am',
 'Temp3pm']
```

```python
analyses =[
         ('MinTemp', 'MaxTemp'),
         ('MaxTemp', 'Humidity9am'),
         ('MaxTemp', 'Pressure3pm'),
         ('MaxTemp', 'Rainfall'),
         ('Rainfall', 'WindSpeed9am'),
         ('Rainfall', 'Humidity9am'),
         ('Rainfall', 'Temp9am')]
```

```python
df_avg = df.groupby(['Year_Month','RainToday'])[filter1].mean().reset_index()
df_avg.head()
```

```python
for analyse in analyses:
    fig = px.scatter(df_avg, x=analyse[0], y=analyse[1],title=f"Scatterplot of {analyse[0], analyse[1]}",
                     trendline="ols",
                     color='RainToday',
                     trendline_color_override="yellow")
    fig.show()
```

# ****Time series (preferably monthly) line chart for numerical variables

```python
for i in filter1:
    fig = px.line(df_avg, x='Year_Month', y=i, title=f"Lineplot of {i}",
                     color='RainToday',
                 )
    fig.show()
```

```python
filter2=[
 'MinTemp',
 'MaxTemp',
 'Rainfall',
 'WindGustSpeed',
 'WindSpeed9am',
 'WindSpeed3pm',
 'Humidity9am',
 'Humidity3pm',
 'Pressure3pm',
 'Pressure9am',
 'Temp9am',
 'Temp3pm']
```

```python
df_avg_month = df.groupby(['Month','RainToday'])[filter1].mean().reset_index()
df_avg_month.head()
```

```python
for i in filter2:
    fig = px.line(df_avg_month, x='Month', y=i, title=f"Lineplot of {i}",
                     color='RainToday',
                 )
    fig.show()
```

# 7- Polar/Radar chart for numerical variables and Wind Direction as the 2nd variable

```python
df_directions = df.groupby('WindGustDir')[filter2].agg('mean').reset_index()
```

```python
import plotly.graph_objects as go

directions = df_directions.WindGustDir.unique()

for i in filter2:
    fig = go.Figure(
        data=go.Scatterpolar(r=df_directions[i], theta=directions, fill='toself')
    )

    fig.update_layout(
        polar=dict(radialaxis=dict(visible=True)),
        title={'text': f"Radarplot of {i}"},
        showlegend=False
    )

    fig.show()
```
