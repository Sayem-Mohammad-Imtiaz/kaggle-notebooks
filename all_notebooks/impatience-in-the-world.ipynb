```python
df = pd.read_csv('../input/most-patient-countries/country-variation-in-patience.csv')
df.tail().style.set_properties(**{'background-color':'purple',
                                     'color': 'white'})
```

```python
df.isnull().sum()
```

```python
#Code by Sandhya S https://www.kaggle.com/sansuthi/sweden-the-most-patient-country

df.Patience = np.round(df.Patience, decimals=3)
df = df.sort_values('Patience', ascending=True).set_index('Country')
df
```

"Many of the world's most patient populations are located in Europe or the English-speaking world, while risk taking is particularly prevalent in Africa and the Middle East. Prosocial preferences are particularly pronounced throughout Asia and relatively weak in sub-Saharan Africa. The various preference measures are also correlated, giving rise to distinct “preference profiles” of groups of countries: Patience and willingness to take risks are one pair of positively correlated preferences, and the prosocial traits of positive reciprocity, altruism, and trust form another grouping."

https://academic.oup.com/qje/article/133/4/1645/5025666

```python
#Code by Sandhya S https://www.kaggle.com/sansuthi/sweden-the-most-patient-country

print(f'Most Patient Country: {df.index[-1]} ({df.Patience[-1]})\
     \nLeast Patient Country: {df.index[0]} ({df.Patience[0]})')
plt.figure(figsize=(13,23))
plot = sns.barplot(df.Patience, df.index)
plt.bar_label(plot.containers[0],size=10)
plt.show()
```

```python
#Code by Sandhya S https://www.kaggle.com/sansuthi/sweden-the-most-patient-country

fig, ax = plt.subplots(1,2, figsize=(14,7))
plt.suptitle(f'Ten Least Patient Countries{" "*45}Ten Most Patient Countries', fontsize=14)
sns.heatmap(df.iloc[:10].drop('Code', axis=1), cmap='crest', 
            annot=True, annot_kws={'size': 14}, ax=ax[0])
sns.heatmap(df.iloc[-11:].drop('Code', axis=1).sort_values('Patience'), 
            annot=True, annot_kws={'size': 14}, ax=ax[1])
plt.show()
```

```python
#Code by Sandhya S https://www.kaggle.com/sansuthi/sweden-the-most-patient-country

map_df = dict(type = 'choropleth', locations = df.index,
                locationmode = 'country names', z = df.Patience, text=df.Code, 
                colorscale= 'turbo', marker = dict(line = dict(color='black',width=0.5)),
                colorbar = {'title':'Patience'})
layout = dict(title = 'Give up today (2012) to benefit more from that in the future?',
              geo = dict(showframe = False, projection = {'type':'natural earth'}))
choromap = go.Figure(data=map_df, layout=layout)
iplot(choromap)
```
