## Big-Mart Data Cleaning

```python
data= pd.read_csv("../input/big-mart-data/bigmart_data.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
## checking for missing values
data.isnull().sum()
```

```python
# imputing missing values with mean for item_weight and mode for outlet_size
data['Item_Weight']=data['Item_Weight'].fillna(data['Item_Weight'].mean())
data['Outlet_Size']=data['Outlet_Size'].fillna(data['Outlet_Size'].mode()[0])
```

```python
# validating
data.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# checking outliers in numerical columns

plt.rcParams['figure.figsize']=(12,4)
plt.style.use('fivethirtyeight')

plt.subplot(1,3,1)
sns.boxplot(data['Item_Weight'])

plt.subplot(1,3,2)
sns.boxplot(data['Item_Visibility'])

plt.subplot(1,3,3)
sns.boxplot(data['Item_MRP'])
```

```python
## Item_visibilty shows outliers
data['Item_Visibility']
```

```python
## removing outliers greater than 0.2
data=data[data['Item_Visibility']<=0.2]

sns.boxplot(data['Item_Visibility'])
```

```python
data['Item_Identifier'].value_counts()
```

```python
## Cleaning the item_identifiers column,the first two characters are same for most of the identifiers
data['Item_Identifier']= data['Item_Identifier'].apply(lambda x: x[0:2])

## validating
data['Item_Identifier'].value_counts()
```

```python
data['Item_Fat_Content'].value_counts()
```

```python
# Low fat is in three different forms and same goes for regular(2), lets replace them

data['Item_Fat_Content']= data['Item_Fat_Content'].replace(('low fat','LF','reg'),('Low Fat','Low Fat','Regular'))

# validating
data['Item_Fat_Content'].value_counts()
```

## Melbourne Housing Data Cleaning

```python
housing= pd.read_csv("../input/melbourne-data/melbourne.csv")
```

```python
housing.head()
```

```python
## Checking the percentage of missing values in all columns
percent_miss_val= round(100*(housing.isnull().sum()/len(housing.index)),2)
percent_miss_val
```

```python
## removing three columns having 33-50% missing values
housing= housing.drop('BuildingArea',axis=1)
housing= housing.drop('YearBuilt',axis=1)
housing= housing.drop('CouncilArea',axis=1)
```

```python
housing.columns
```

```python
## remmoving missing value rows from price columns
housing= housing[~np.isnan(housing['Price'])]
```

```python
## remmoving missing value rows from Landsize columns
housing= housing[~np.isnan(housing['Landsize'])]
```

```python
## Re-checking the missing percentage
percent_miss_val= round(100*(housing.isnull().sum()/len(housing.index)),2)
percent_miss_val
```

```python
## describing the values for Lat and Long
housing[['Lattitude','Longtitude']].describe()
```

```python
## we can see that there is a minute difference between the min, 25%,50%,75% and max values, therfore we'll use mean fucnction to impute the missing values.
housing['Lattitude']= housing['Lattitude'].fillna(housing['Lattitude'].mean())
housing['Longtitude'].fillna(housing['Longtitude'].mean(),inplace=True)
```

```python
## Re-checking the missing percentage
percent_miss_val= round(100*(housing.isnull().sum()/len(housing.index)),2)
percent_miss_val
```

```python
## checking values of bathroom column
housing['Bathroom'].value_counts()
```

```python
## Imputing values into the categorical columns using mode
housing['Bathroom'].fillna(housing['Bathroom'].mode()[0],inplace=True)
housing['Car'].fillna(housing['Car'].mode()[0],inplace=True)
```

```python
## lets check the final missing data
housing.isnull().sum().sum()
```
