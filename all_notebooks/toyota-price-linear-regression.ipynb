```python
df=pd.get_dummies(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
```

```python
x=df.iloc[:,1:]
y=df.iloc[:,0]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
reg=LinearRegression().fit(x_train,y_train)
```

```python
reg.score(x_test,y_test)
```

```python
np.corrcoef(reg.predict(x_test),y_test)
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x=reg.predict(x_test),y=y_test,palette='crest')
```

```python
regressor=DecisionTreeRegressor(random_state=0).fit(x_train, y_train)
```

```python
regressor.score(x_test,y_test)
```

```python
np.corrcoef(regressor.predict(x_test),y_test)
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x=regressor.predict(x_test),y=y_test,palette='crest')
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR
sc=StandardScaler()
scaled_x=sc.fit_transform(x)
x_train,x_test,y_train,y_test=train_test_split(scaled_x,y,test_size=0.2)
svm=SVR(kernel='linear').fit(x_train,y_train)
```

```python
svm.score(x_test,y_test)
```

```python
np.corrcoef(svm.predict(x_test),y_test)
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(x=svm.predict(x_test),y=y_test,palette='crest')
```
