```python
poke_df = pd.read_csv('../input/Pokemon.csv')
```

```python
poke_gen1_raw_df = poke_df[poke_df['Generation'] == 1]
poke_gen1_df = poke_gen1_raw_df[~poke_gen1_raw_df['Name'].str.contains('Mega')]
```

```python
#credit for function goes to Peter from stackoverflow
#http://stackoverflow.com/questions/32791911/fast-calculation-of-pareto-front-in-python

def is_pareto_efficient(costs):
    """
    :param costs: An (n_points, n_costs) array
    :return: A (n_points, ) boolean array, indicating whether each point is Pareto efficient
    """
    is_efficient = np.ones(costs.shape[0], dtype = bool)
    for i, c in enumerate(costs):
        if is_efficient[i]:
            # Remove dominated points
            is_efficient[is_efficient] = np.any(costs[is_efficient]<=c, axis=1)  
    return is_efficient
```

```python
#here are the attributes we wish to construct our pareto frontier with
stat_list = ['HP','Attack', 'Defense', 'Sp. Def', 'Sp. Atk', 'Speed']

#the above function asks for a cost vector, so we multiply our stats by -1
pareto_indexer = is_pareto_efficient(np.array(poke_gen1_df[stat_list])*-1)
```

```python
pareto_frontier_df = poke_gen1_df[pareto_indexer]
```

### Below is a list of Pokemon on the Pareto frontier. They are sorted by their total score. 

```python
pareto_frontier_df.sort_values('Total', ascending=False)
```

If you were creating a lineup, it would not make sense to choose a Pokemon that was not on this list. All 106 Pokemon not on this list represent **strictly dominated** alternatives. Pokemon A is considered strictly dominated if there exists Pokemon B that outperforms Pokemon A in all categories.

To illustrate the Pareto frontier, we’re going to compare the Pokemon on this list with the highest and lowest scores - Mewtwo and Onix respectively. Even though Mewtwo’s overall score is much better than Onix’s, Mewtwo does not strictly dominate Onix because Onix is better than Mewtwo in at least one category - in this case defense.
