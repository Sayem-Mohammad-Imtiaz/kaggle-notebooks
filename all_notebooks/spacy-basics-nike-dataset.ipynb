```python
import pandas as pd
import numpy as np
import spacy
import seaborn as sns
import random
```

```python
nlp = spacy.load('en')
```

```python
tweets = pd.read_csv("../input/justdoit_tweets_2018_09_07_2.csv")
```

```python
tweets.shape
```

Picking some Random Tweets from the dataset.

```python
text = tweets.tweet_full_text[random.sample(range(1,100),10)]
text
```

```python
combined = str(text)
doc = nlp(combined)
```

```python
for token in doc:
    print(token)
```

```python
for token in doc:
    print(token.text,token.pos_)
```

Extracting Sentences

```python
sents = list(doc.sents)
sents
```

Named Entity Recognition

```python
for entity in doc.ents:
    print(entity.text,entity.label_)
```

```python
#Entity style Display
spacy.displacy.render(doc, style='ent',jupyter=True)
```

```python
#Lemmatization
for token in doc:
    print(token.text,token.lemma_)
```

```python
#Dependency Parser style Display
spacy.displacy.render(doc, style='dep',jupyter=True)
```
