```python
import pandas as pd
import numpy as np
import holoviews as hv
import seaborn as sns
import matplotlib.pyplot as plt
hv.extension('bokeh')
```

```python
data=pd.read_csv("/kaggle/input/okcupid-profiles/okcupid_profiles.csv")
data.head()
```

```python
data.dtypes
```

```python
data["sex"].unique()
```

```python
## Make a category of Status
data["sex"]=data["sex"].astype("category")
data["sex"].cat.set_categories(["m","f"], inplace=True)
```

```python
pd.pivot_table(data=data, index=data.sex, values="income", aggfunc=np.mean)
```

on average, the average age for both men and women is 32, men are taller than women and receive on average a twice higher salary

```python
pd.pivot_table(data=data, index=data.education, values=["age", "income"],aggfunc=np.mean)
```

```python
pd.pivot_table(data=data, index=["education", "sex"],aggfunc=np.mean)
```

```python
pd.pivot_table(data=data, index=["smokes", "drinks"], values="income")
```

```python
pd.pivot_table(data=data, index=["smokes", "drinks"], values=["income"])
```

```python
pd.pivot_table(data=data, index="education", values=["income"], aggfunc=[np.mean,len], fill_value=0)
```

```python
pd.pivot_table(data=data, 
               index=["status", "sex"], 
               values=["income", "age"],
               aggfunc=np.mean, 
               fill_value=0, 
               margins=True)
```

on Average the age is 32y.o and the average income is around 20 000â‚¬

```python
pd.pivot_table(data=data, 
               index=["status", "sex"], 
               values=["income", "age"],
               aggfunc={"income": np.mean, "age": np.median}, 
               fill_value=0, 
               margins=True)
```

```python
table=pd.pivot_table(data=data, 
               index=["status", "sex"], 
               values=["income", "age"],
               columns="orientation",
               aggfunc={"income": np.mean, "age": np.median}, 
               fill_value=0, 
               margins=True)
table
```

```python
table=pd.pivot_table(data=data, 
               index=["status", "sex"], 
               values=["income", "age"],
               columns="orientation",
               aggfunc=[len, np.mean],
               fill_value=0, 
               margins=True)
table
```

```python
table.query('status==["available"]')
```

```python
table.query('status==["available", "married"]')
```
