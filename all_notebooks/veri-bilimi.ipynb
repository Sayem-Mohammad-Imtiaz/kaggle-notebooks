```python
# NESTED FUNCTION

def square():
    """return square of vlaue"""
    def add():
        """add two local variable"""
        x = 2
        y = 3
        z = x+y
        return z
    return add()**2
print(square())
```

```python
# DEFAULT and FLEXIBLE ARGUMENTS

# DEFAULT
print("Default")
def f(a,b = 3,c=2):
    y = a+b+c
    return y
print(f(5))
print(f(5,20,100))

# FLEXIBLE
print("\nFLEXIBLE")
def f(*args):
    for i in args:
        print(i)
f(1)
print("")
f(4,6,3,77,9)
```

**EXPLORATORY DATA ANALYSIS**

value_counts(): Frekans sayıları

outliers: Verilerin geri kalanından önemli ölçüde daha yüksek veya daha düşük olan değer

* Diyelim ki, 75'teki değer Q3 ve% 25'teki değer Q1'dir.
* Aykırı Q1 - 1.5'ten (Q3-Q1) küçük ve Q3 + 1.5'den (Q3-Q1) büyük. (Q3-Q1) = IQR
* Define () metodunu kullanacağız. Tarif yöntemi içerir:
* count: girişlerin sayısı
> > * ortalama: girişlerin ortalaması
* std: standart sapma
* min: minimum giriş
* % 25: ilk miktar
* % 50: ortanca veya ikinci kuantil
* % 75: üçüncü miktar
* max: maksimum giriş

RESAMPLING PANDAS TIME SERIES

    Resampling: statistical method over different time intervals
        Needs string to specify frequency like "M" = month or "A" = year
    Downsampling: reduce date time rows to slower frequency like from daily to weekly
    Upsampling: increase date time rows to faster frequency like from daily to hourly
    Interpolate: Interpolate values according to different methods like ‘linear’, ‘time’ or index’
    
    
    YEDEK PANDAS ZAMAN SERİSİ

    Örnekleme: farklı zaman aralıklarında istatistiksel yöntem
        Dize "M" = ay veya "A" = yıl gibi bir sıklık belirtmesi gerekiyor
    Altörnekleme: tarih saat satırlarını günlükten haftaya gibi daha yavaş bir frekansa düşürün
    Örnekleme: tarih saat satırlarını günlükten saat başı gibi daha hızlı bir sıklığa yükseltin
    Enterpolasyon: Değerleri 'doğrusal', 'zaman' veya dizin 'gibi farklı yöntemlere göre enterpolasyon yapın.

```python
data2.resample("A").mean()
```

```python
data2.resample("M").mean()
```

```python
# In real life (data is real. Not created from us like data2) we can solve this problem with interpolate
# We can interpolete from first value
data2.resample("M").first().interpolate("linear")
```

```python
# Or we can interpolate with mean()
data2.resample("M").mean().interpolate("linear")
```
