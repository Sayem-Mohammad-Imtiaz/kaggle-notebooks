```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from pandas.core.reshape.pivot import pivot_table
```

```python
FULL_PATH = "/kaggle/input/accidents-kenya/accidents-database-.csv"
# Reading the file into Pandas
data = pd.read_csv(FULL_PATH)
```

```python
# Removing double header & NULL Values
data_set = data[data["NO."].str.contains("NO.") == False]
data_set.dropna()
# Converting numbers into INT
data_init = data_set.copy()
data_init["NO."] = data_init['NO.'].astype("int")
```

```python
clean_set = data_init[['TIME 24 HOURS', 'COUNTY', 'ROAD', 'GENDER', 'AGE', 'NO.', 'Date DD/MM/YYYY']]
print(clean_set.info())
```

```python
# Ploting Counties with high number to victims
victim_by_county = clean_set[['COUNTY', 'NO.']]
pivot_county_victims = pivot_table(victim_by_county, values="NO.", index="COUNTY")
pivot_county_victims.plot.bar()
plt.show()
```

```python
# Ploting Roads with high number to victims where Casualty > 2
victim_by_road = clean_set[clean_set["NO."]  > 2]
pivot_county_road = pivot_table(victim_by_road, values="NO.", index="ROAD")

pivot_county_road.plot.bar()
plt.show()
```

```python
# Determining the time of day that the accident happened. 
remove_char = ["HRS", "UNKNOWN", "hrs", "NAROK"]
rm_string = '|'.join(remove_char)
accident_time = clean_set[clean_set["TIME 24 HOURS"].str.contains(rm_string) == False]
cl_accident_time = accident_time[["TIME 24 HOURS", "NO."]]

pivot_accident_time = cl_accident_time.melt(id_vars="TIME 24 HOURS")

pivot_accident_time.plot(x="TIME 24 HOURS")
plt.show()
```
