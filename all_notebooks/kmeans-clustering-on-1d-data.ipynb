```python
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
!pip install kneed
```

```python
df = pd.read_csv('../input/utk-face-extracted-info/extracted_info.csv')
df.dropna(inplace=True)
df = df[df['Age'] <= 100]
        
df = df.reset_index()
df = df.drop(columns = {'DateTime', 'Unnamed: 0', 'index'})
df = df.dropna()

num_classes = len(df['Age'].unique())
df.head(3)
```

```python
sum_squared_dist = []
for k in range(7, 30):
    km = KMeans(n_clusters=k, random_state=0)
    km = km.fit(df[['Age']])
    sum_squared_dist.append(km.inertia_)
```

```python
plt.figure(figsize=(22, 9))
plt.ylim(0, 280000)

from kneed import KneeLocator
x = np.array(range(7, 7 + len(sum_squared_dist)))
kn = KneeLocator(x, sum_squared_dist, curve='convex', direction='decreasing')
print(kn.knee)

plt.xlabel('number of clusters k')
plt.ylabel('Sum of squared distances')
plt.plot(x, sum_squared_dist)
plt.vlines(kn.knee, 0, 270000, linestyles='dashed', color='green', label='elbow point')
plt.show()
```

```python
km = KMeans(n_clusters=14, random_state=0)
km = km.fit(df[['Age']])
df['clusters'] = km.predict(df[['Age']])
```
