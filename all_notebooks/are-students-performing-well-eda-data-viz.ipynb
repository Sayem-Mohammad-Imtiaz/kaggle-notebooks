```python
import pandas as pd
import numpy as np
df = pd.read_csv("../input/students-performance-in-exams/StudentsPerformance.csv")
df.head(10).style.set_properties(**{'background-color': 'black',                                                   
                                    'color': 'lawngreen',                       
                                    'border-color': 'white'})
```

```python
from sklearn.preprocessing import LabelEncoder

for c in df.columns:
    if df[c].dtype=='object': 
        lbl = LabelEncoder()
        lbl.fit(list(df[c].values))
        df[c] = lbl.transform(df[c].values)
        
df.head(10).style.background_gradient(cmap = 'hot') 
```

```python
df.describe().T.style.bar(subset=['mean'], color='#205ff2')\
                            .background_gradient(subset=['std'], cmap='Reds')\
                             .background_gradient(subset=['50%'], cmap='twilight')
```

```python
sns.catplot(x="race/ethnicity", y="math score",
                hue="gender", col="lunch",
                data=df, kind="bar",
                height=4, aspect=.7, saturation = 5);
```

<h1 style="color:Olive;">Meal is important, hence if we look closely when lunch is provided math score is comparatively higher than lunch is absent.</h1>

```python
sns.catplot(x="race/ethnicity", y="math score",
                hue="gender", col="parental level of education",
                data=df, kind="bar", saturation = 10,
                height=4, aspect=.7);
```

<h1 style="color:Tomato;">Math scores based on parental level of education and race/ethnicity, For male math score is comparatively higher than female, an exception can be seen in when parental level of education is 3(master's degree) with race/ethnicity is 1(group B) where math scores of female is much higher than male.</h1>

```python
sns.catplot(x="race/ethnicity", y="reading score",
                hue="gender", col="parental level of education",
                data=df, kind="bar",
                height=5, aspect=.7, saturation = 1);
```

<h1 style="color:Maroon;">Reading scores based on parental level of education and race/ethnicity, For female reading score is comparatively higher than male in all aspects of parental level of education.</h1>

```python
sns.catplot(x="race/ethnicity", y="writing score",
                hue="gender", col="parental level of education",
                data=df, kind="bar",
                height=5, aspect=.7, saturation = 1);
```
