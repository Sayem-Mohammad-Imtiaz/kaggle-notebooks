# # Importing libraries

```python
import pandas as pd
import nltk
nltk.download("stopwords")
nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
nltk.download("maxent_ne_chunker")
nltk.download("words")
import re
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

**Inputting string data for extraction**

```python
string = """
Hello I am a dataScience learner as u have seen my name from kaggle profile "Faraz Shaikh"
My phone number is 3153644364 and 3042780709
Email : farazshaikh581@gmail.com and for urgent contact : farazshaikh2018@outlook.com
You can see my work on github also: github.com/farazshaikh581

"""
```

**Defining a function for Phone numbers extracion using Regular Expression**

```python
def extract_phone_numbers(string):
    r = re.compile(r'(\d{3}[-\.\s]??\d{3}[-\.\s]??\d{4}|\(\d{3}\)\s*\d{3}[-\.\s]??\d{4}|\d{3}[-\.\s]??\d{4})')
    phone_numbers = r.findall(string)
    return [re.sub(r'\D', '', number) for number in phone_numbers]
```

**Defining a function for Email Address extracion using Regular Expression**

```python
def extract_email_addresses(string):
    r = re.compile(r'[\w\.-]+@[\w\.-]+')
    return r.findall(string)
```

**Defining function to separate data in chunks.**

```python
def ie_preprocess(document):
    document = ' '.join([i for i in document.split() if i not in stop])
    sentences = nltk.sent_tokenize(document)
    sentences = [nltk.word_tokenize(sent) for sent in sentences]
    sentences = [nltk.pos_tag(sent) for sent in sentences]
    return sentences
```

**Defining a function for Names extracion using Regular Expression**

```python
def extract_names(document):
    names = []
    sentences = ie_preprocess(document)
    for tagged_sentence in sentences:
        for chunk in nltk.ne_chunk(tagged_sentence):
            if type(chunk) == nltk.tree.Tree:
                if chunk.label() == 'PERSON':
                    names.append(' '.join([c[0] for c in chunk]))
    return names
```

**Finally, a condition to execute above fuctions**

```python
if __name__ == '__main__':
    numbers = extract_phone_numbers(string)
    emails = extract_email_addresses(string)
    names = extract_names(string)
```
