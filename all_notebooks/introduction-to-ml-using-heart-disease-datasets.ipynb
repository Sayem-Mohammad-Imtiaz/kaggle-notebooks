## Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.array([0,90])
b = np.array([1,50])

plt.plot(a,b)
plt.show()
```

```python
a = np.array([2,7,8,20])

plt.plot(a,linestyle='dashed')
plt.show()
```

```python
a = np.array([2,7,8,20])

plt.plot(a,'o:y')
plt.show()
```

```python
a = np.array([2,7,8,20])
plt.plot(a,marker='o', ms = 15, mec = 'y',mfc = 'r')
plt.show()
```

```python
a = np.array([0,40])
b = np.array([1,20])

plt.plot(a)
plt.plot(b)
plt.xlabel('clubs')
plt.ylabel('players')
plt.title("Goals" )
plt.grid()
plt.show()
```

```python
x = np.random.normal(189, 28, 150)
plt.hist(x)
plt.show() 
```

```python
y = np.array([3, 2, 5, 8])
mylabels = ["Dybala", "messi", "CR7", "Haaland"]
myexplode = [0.2, 0, 0, 0]
mycolors = ["red", "hotpink", "b", "g"]
plt.pie(y, labels = mylabels,startangle = 90, explode = myexplode, shadow = True, colors = mycolors)
plt.legend(title='Players')
plt.show() 
```
