```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('../input/BR_eleitorado_2016_municipio.csv', delimiter=',')
df.head(5)
```

# Questão 1

Construa um gráfico de violino que compara o número de eleitores do gênero masculino e gênero feminino das 15 cidades com o maior número de eleitores do estado da Bahia.

```python
estado_ba = df.loc[df['uf'] == 'BA'].nlargest(15,'total_eleitores')
plt.violinplot([estado_ba['gen_feminino'],estado_ba['gen_masculino']],showmeans=True, showmedians=True,widths = 1.0, bw_method=0.1) #default
plt.xticks([0,1,2], ('', 'Feminino','Masculino'))

plt.show()
```

# Questão 2

Construa um segundo gráfico de violino que compara o número de eleitores com idades entre 18-20 anos, 21-24 anos, 70-79 anos e eleitores com mais de 79 anos, para as 15 cidades com o maior número de eleitores do estado de Minas Gerais.

```python
estado_mg = df.loc[df['uf'] == 'MG' ].nlargest(15,'total_eleitores')

plt.violinplot([estado_mg['f_18_20'],estado_mg['f_21_24'],estado_mg['f_70_79'],estado_mg['f_sup_79']],showmeans=True, showmedians=True,widths = 1.0, bw_method=0.1) #default
plt.xticks([0,1,2,3,4], ('', '18_20','21_24', '70_79','f_sup_79'))

plt.show()
```

# Questão 3

Para cada uma das questões anteriores construa um gráfico de caixa correspondente. (total de dois gráficos de caixa)

```python
df_ba = pd.DataFrame(estado_ba[['gen_masculino','gen_feminino']])
df_ba.plot.box()

df_mg = pd.DataFrame(estado_mg[['f_18_20','f_21_24','f_70_79','f_sup_79']])
df_mg.plot.box()
```
