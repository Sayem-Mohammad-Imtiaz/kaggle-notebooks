```python
df = data[(data["Type"]=="1 RK Apartment") |(data["Type"]=="3 BHK Apartment")|(data["Type"]=='2 BHK Apartment') |(data["Type"]=='1 BHK Apartment')]
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=df,x="Locality",y="Rent/Month",marker='o',hue="Furnishing")
plt.xticks(rotation=90)
plt.show()
```

```python
df['Build_up_area(sq.ft)']=df['Build_up_area(sq.ft)'].str.split('sq.ft', expand=True)
```

```python
df['Build_up_area(sq.ft)']=df['Build_up_area(sq.ft)'].astype(float)
```

```python
plt.figure(figsize=(20,7))
sns.lineplot(data=df,x="Locality",y="Build_up_area(sq.ft)" , marker='+')
plt.xticks(rotation=90)
plt.show()
```
