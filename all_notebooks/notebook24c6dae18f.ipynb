```python
# (WR) = (v ÷ (v+m)) × R + (m ÷ (v+m)) × C where:

# R = average for the movie (mean) = (Rating)
# v = number of votes for the movie = (votes)
# m = minimum votes required to be listed in the Top 250 (currently 15000)
# C = the mean vote across the whole report (currently 7.0)
```

```python
m = 15000
```

```python
C= new_df[new_df['num_voted_users'] > 15000]['imdb_score'].mean()
```

```python
def cal_weighted_rating(row, m=m, C=C):
    v = row['num_voted_users']
    R = row['imdb_score']
    wr = ((v/(v+m)) * R) + ((m/(v+m)) * C)
    return wr
```

```python
new_df['wr'] = new_df.apply(cal_weighted_rating, axis=1)
```
