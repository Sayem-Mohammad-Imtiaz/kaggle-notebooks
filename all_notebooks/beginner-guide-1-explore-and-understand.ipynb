** What is The Iris Dataset**

Iris dataset includes petal and sepal length and width of each flower  



and 3 diffrent types of iris species(50 records of each) which are:



*Virginica*


*Setosa*


*Versicolor*


So overall dataset contains 5 columns and 150 rows.

**How to check that?**

first we need to read IRIS.csv file to be able to work with it.

```python
data = pd.read_csv('/kaggle/input/iris-flower-dataset/IRIS.csv')
```

Now while we defined .csv file as 'data' simply use .shape tuple that gives dimensions of dataset.

```python
data.shape
```

Next step is to take a look at dataset.

```python
data.head()
```

 Seems like I've been right this time! IRIS.csv file contains 5 columns sepal_length,sepal_width,petal_length,petal_width,species.
 
 Now let's make sure dataset contains 50 records of each type of flower and it has no null values.

```python
data.groupby('species').size()
```

```python
data.info()
```

**how to distinguish one from another?**

Well there's many ways to do that but in this case easiest solution is to compare mean length of each specie one another.

```python
data.groupby('species').mean()
```

**Data Visualization**

Here's very simple way to visualize data using all parameters of data.

```python
import seaborn as sns
sns.pairplot(data,hue ='species')
```
