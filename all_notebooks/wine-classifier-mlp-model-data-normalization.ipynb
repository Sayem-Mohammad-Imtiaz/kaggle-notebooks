```python
import pandas as pd
base = pd.read_csv('../input/wine-dataset/wine_dataset.csv')
```

```python
# Substituindo os valores do atributo classe ('style') para '0' (tinto/vermelho) e '1' (branco)
base['style'] = base['style'].replace('red', 0)
base['style'] = base['style'].replace('white', 1)
```

```python
# 'y' representa os valores de saída considerando todas as amostas
y = base['style']

# 'x' representa os registros com seus respectivos atributos, com execeção do atributo classe ('style')
x = base.drop('style', axis = 1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Cálculo da média e desvio padrão dos dados
scaler.fit(x)

# Normalização dos dados
x = scaler.transform(x)
```

```python
from sklearn.model_selection import train_test_split

# Separação dos dados treino e teste (30 por cento para teste)
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.3)
```

```python
from sklearn.neural_network import MLPClassifier

# Construção da Rede Neural MLP
mlp_model = MLPClassifier(solver='adam', hidden_layer_sizes = {90,50},random_state=1, max_iter=300, activation='logistic')

# Treinamento da rede
mlp_model.fit(x_treino, y_treino)

# Previsão das saídas considerando os registros de teste
previsao = mlp_model.predict(x_teste)
```

```python
from sklearn.metrics import accuracy_score

# Desempenho (acurácia) do modelo(MLP) durante a classificação do registros de teste (0.9974358974358974 ~ 0.997948717948718)
print(accuracy_score(y_teste, previsao))
```
