Milline oli keskmine rahulolu hinne kõikide riikide lõikes?

```python
keskmine_rahulolu_hinne = df["Happiness Score"].mean()
print("Keskmine rahulolu hinne:", df["Happiness Score"].mean())
```

Millistel riikidel oli rahulolu hinne üle keskmise?

```python
keskmine_rahulolu_hinne = df["Happiness Score"].mean()
üle_keskmise = df[df["Happiness Score"] > keskmine_rahulolu_hinne]
```

```python
üle_keskmise
```

Teen uue tabeli, milles on näha riik (mille hinne oli üle keskmise), selle hinne ja piirkond, kus asub.

```python
üle_keskmise = pd.DataFrame({"Region" : üle_keskmise["Region"],
             "Happiness_score" : üle_keskmise["Happiness Score"],
                            "Country" : üle_keskmise["Country"]})
üle_keskmise
```

Leian, millised olid erinevate piirkondade keskmised rahulolu hinded.

```python
üle_keskmise.groupby(["Region"])["Happiness_score"].mean()
```

Siit saab järeldada, et Põhja-Ameerikas, Austraalias ja Uus-Meremaal on kõige kõrgem rahulolu hinne. Kõige madalam rahulolu hinnang on Mustas Aafrikas.

Histogramm - näitab, mitu riiki sai mis hinde (nt hinde 6 sai 7 riiki)

```python
üle_keskmise.Happiness_score.plot.hist(bins=20, grid=True, rwidth=0.8);
```

Histogramm näitab, et üle hinde 7.5 on saanud ainult 4 riiki, mis on väga vähe. Enamus riikide, kes on saanud üle keskmise, hinne jääb pigem sinna 5-6 vahele.

Hajusdiagramm - kuidas on seotud rahulolu hinne ja inimeste oodatav eluiga?
