```python
dfTemp=pd.read_csv('../input/queryfetchtime/QueryFetchTime.csv')
```

```python
dfTemp.info()
dfTemp.head(20)
```

```python
df=dfTemp[["LineNo","Dated","Timed","Thread","TimeTaken"]]
#df.sort_values(by=["TimeTaken"], axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=True)
df.info()
df.describe()
```

```python
n, bins, patches = plt.hist(x=df[["TimeTaken"]].to_numpy(), bins='auto', color='#0504aa',
                            alpha=0.7,rwidth=1.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Time Taken')
plt.ylabel('Counts')
plt.title('Query Response Times')
#plt.xticks([1000])
maxfreq = n.max()
# Set a clean upper y-axis limit.
plt.ylim(ymax=np.ceil(maxfreq / 10) * 10 if maxfreq % 10 else maxfreq + 10)
```
