```python
#importing pandas
import pandas as pd
#importing matplotlib.pyplot as plt
import matplotlib.pyplot as plt
#importing LinearRegression and LogisticRegression from sklearn.linear_model 
from sklearn.linear_model import LinearRegression , LogisticRegression
#importing train_test_split from sklearn.model_selection 
from sklearn.model_selection import train_test_split
#importing seaborn as sns
import seaborn as sns
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

In this stage we will use pandas for import data by method called **read_csv()** to read file located in path **"/kaggle/input/sallary/Salary.csv"** in the first line of the code. 

The second and third lines in the code we use **iloc()** method to split the data in **x** and **y**.

how **iloc()** works?

simply iloc[:,index of start column: index of end column]

to split Feature it will be take all column without the least column which is the target column and will stored in **x**.

to split the target we will split the least column by use the index of the clumn by '-1' and will stored in **y**.

we will use **train_test_split** to split the data in 4 variables: **x_train,x_test,y_train,y_test** by pass **x** and **y** and tuning the data by split 30% of the data to make it as testing data.

```python
data = pd.read_csv('/kaggle/input/salary/Salary.csv')
x = data.iloc[:, :-1].values    # Features => Years of experience => Independent Variable
y = data.iloc[:, -1].values     # Target => Salary => Dependent Variable
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0) #spliting data
```

As a general aggrement in the full note book we will use some ways to implementation rebeatlly:

1. We will creat an object and we will named it be the type of Regression like linear Regression will calles lenReg, len for Linear and Reg for Regression 
2. we will fit the x_train, y_train by the Regression model. 
3. to decrease the number of variables and optimize the memory usage, insted of store predictions in y_pred or predictions as usual we will call it in the sns.displot and use it.

```python
#LinearRegressionModel
lenReg = LinearRegression()
lenReg.fit(x_train, y_train)
sns.distplot(lenReg.predict(x_test)-y_test) #the predictions = lenReg.predict(x_test)
```

```python
#Linear Regression plot
plt.scatter(x_train, y_train, color='black')
plt.plot(x_train, lenReg.predict(x_train))
```

```python
#LogesticReggression
logReg=LogisticRegression()
logReg.fit(x_train,y_train)
sns.distplot(logReg.predict(x_test)-y_test) #the predictions = logReg.predict(x_test)
```

```python
#Logistic Regression plot
plt.scatter(x_train, y_train, color='black')
plt.plot(x_train, logReg.predict(x_train))
```
