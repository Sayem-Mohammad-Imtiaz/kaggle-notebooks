```python
#Importing all the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.linear_model import BayesianRidge
from sklearn.ensemble import RandomForestRegressor
from lightgbm import LGBMRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import math
```

```python
#Read the dataset file in dataframe
df = pd.read_csv('../input/flight-take-off-data-jfk-airport/M1_final.csv')
df
```

```python
df.info()
#dropping rown containg NaN values
df = df.dropna()
```

```python
df.info()
```

```python
df['Dew Point'].unique()
```

```python
#Replacing '\xa0' from 'Dew Point'
df = df.replace('\xa0','', regex=True)
df['Dew Point'].unique()
```

```python
#Converting data of the column 'Dew Point' to integer
s = pd.to_numeric(df['Dew Point'])
df['Dew Point'] = s
df.info()
```

```python
df['Dew Point'].unique()
```

```python
col = list(df.columns)
dic = {}
for i in range(0,len(col)):
    dic[col[i]] = i
dic
#Renaming name of the columns by integers
df = df.rename(columns = dic)
```

```python
#Getting the feature columns
X = df.iloc[:, :22]
#Getting the label columns
y = df.iloc[:,22]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
X_train
#Concatenate features and labels along columns i.e axis =1
df_train = pd.concat([X_train, y_train], ignore_index=True, sort=False, axis =1)
df_test = pd.concat([X_test, y_test], ignore_index=True, sort=False, axis =1)

# Saving preprocessed data
# We can use the same splitted data for Label encoding and One Hot Encoding
df_train.to_csv('train.csv', index=False)
df_test.to_csv('test.csv', index=False)
```

```python
col = list(df.columns)
df = pd.read_csv('train.csv')
df1 = pd.read_csv('test.csv')
#doing label encoding with the required columns
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
# Encode labels in column other than integer and float.
for i in range(0,len(col)):
    if df[col[i]].dtype == 'O':
        df[col[i]]= label_encoder.fit_transform(df[col[i]])
    if df1[col[i]].dtype == 'O':
        df1[col[i]]= label_encoder.fit_transform(df1[col[i]])
df
```

```python
X_train = df.iloc[:,:22]
y_train = df.iloc[:,22]
X_test = df1.iloc[:,:22]
y_test = df1.iloc[:,22]
```

```python
MAE = []
RMSE = []
R2 = []
MSE = []

def Score_diplay(y_pred, y_test):
    print("MAE: {}".format(mean_absolute_error(y_pred, y_test)))
    MAE.append(mean_absolute_error(y_pred, y_test))
    print("RMSE: {}".format(math.sqrt(mean_squared_error(y_pred, y_test))))
    RMSE.append(math.sqrt(mean_squared_error(y_pred, y_test)))
    print("r2_score: {}".format(r2_score(y_pred, y_test)))
    R2.append(r2_score(y_pred, y_test))
    print('Coefficients: \n', regressor.coef_)
    # The mean squared error
    print("Mean squared error: %.2f" % np.mean((y_pred - y_test) ** 2))
    MSE.append(np.mean((y_pred - y_test) ** 2))
    print('\n')
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
print('Linear Regression:\n')
Score_diplay(y_pred, y_test)

lasreg=Lasso(alpha=0.05, normalize=True)
lasreg.fit(X_train,y_train)
y_pred=lasreg.predict(X_test)

print('Lasso Regression:\n')
Score_diplay(y_pred, y_test)

ridgereg=Ridge(alpha=0.05, normalize=True)
ridgereg.fit(X_train,y_train)
y_pred=ridgereg.predict(X_test)

print('Ridge Regression:\n')
Score_diplay(y_pred, y_test)

rfreg = RandomForestRegressor()
rfreg.fit(X_train,y_train)
y_pred = rfreg.predict(X_test)

print('Random Forest Regression:\n')
Score_diplay(y_pred, y_test)

knnreg = KNeighborsRegressor(n_neighbors = 200)
knnreg.fit(X_train, y_train)
y_pred = knnreg.predict(X_test)

print('K-Neighbours Regression:\n')
Score_diplay(y_pred, y_test)

lgbm=LGBMRegressor()
lgbm.fit(X_train,y_train)
y_pred=lgbm.predict(X_test)


print('LGBM Regression:\n')
Score_diplay(y_pred, y_test)

bayreg=BayesianRidge()
bayreg.fit(X_train,y_train)
y_pred=bayreg.predict(X_test)

print('Gaussian Naive Bayes:\n')
Score_diplay(y_pred, y_test)

svreg = SVR()
svreg.fit(X_train, y_train)
y_pred = svreg.predict(X_test)

print('SVM:\n')
Score_diplay(y_pred, y_test)

LE_MAE = MAE
LE_RMSE = RMSE
LE_R2 = R2
LE_MSE = MSE
```
