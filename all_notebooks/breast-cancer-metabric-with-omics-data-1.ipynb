# Load data and join clinical to Omics data

```python
df1 = pd.read_csv('../input/breast-cancer-metabric/Breast Cancer METABRIC.csv', index_col = 0)

print(f'METABRIC Dataset Shape: {df1.shape}')
print(f'METABRIC Dataset Memory Usage: {df1.memory_usage().sum() / 1024 ** 2:.2f} MB')

df2 = pd.read_csv('/kaggle/input/breast-cancer-omics-bulk-data/METABRIC.txt', sep = '\t', index_col = 0)
df2 = df2.transpose()
print('Omics data shape', df2.shape)

i2 = [s.replace('BRCA-METABRIC-S1-','') for s in df2.index ]
print('number of common ids:', len(set(i2) & set(df1.index) ) )
df2.index = i2
df = df1.join(df2, how = 'inner')
print('Joined data shape', df2.shape)

df
```

```python
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
pca = PCA
r = pca().fit_transform(X = df[df2.columns].values)
f = 'Pam50 + Claudin-low subtype'
plt.figure(figsize = (20,10))
sns.scatterplot( x=r[:,0], y=r[:,1], hue = df[f] )
plt.title('PCA for Omics data colored by Pam50 groups')
plt.show()
```
