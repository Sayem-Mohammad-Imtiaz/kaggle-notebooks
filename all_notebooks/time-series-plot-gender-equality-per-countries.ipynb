```python
table2 = pd.pivot_table(df, values='value', index=['category', 'year'], columns=['country_or_area'])
table2.head()
# similar to my other kernel : https://www.kaggle.com/jihyeseo/cooling-house-warms-up-the-earth
```

```python
countryNames = table2.columns.values
```

```python
categorynames = df.category.unique().tolist()
```

```python
categorynames
```

```python
def cat_countries_plot3(nameOfCategory, namesOfCountries):
    data = table2.loc[nameOfCategory]
    data.plot( y = namesOfCountries)
    plt.legend(namesOfCountries)
    plt.title(nameOfCategory)
    
cat_countries_plot3(categorynames[4],countryNames[:10])
```

```python
cat_countries_plot3(categorynames[10],countryNames[:100])
```
