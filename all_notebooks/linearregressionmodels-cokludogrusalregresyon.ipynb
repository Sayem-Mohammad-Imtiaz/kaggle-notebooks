```python
import pandas as pd
import numpy as np
import seaborn as sns
df = pd.read_csv('../input/advertising-data/Advertising.csv')
df = df.iloc[:,1:len(df)]
df.head()
```

```python
X = df.drop('Sales',axis = 1)
y = df[['Sales']]
```

```python
X.head()
```

```python
y.head()
```

##### Statsmodels ile model kurmak

```python
import statsmodels.api as sm
```

```python
lm = sm.OLS(y,X)
```

```python
model = lm.fit()
```

```python
model.summary()
```

* R-squared : bagimsiz degiskenlerin bagimli degiskenleri aciklama yuzdesidir.
* Adj. R-squared : daha saglikli aciklanabilirlik
* F-statics : modelin anlamliligini test etmek icin kullanabilir

##### Sckit learn ile model kurmak

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
```

```python
model = lm.fit(X,y)
print(model.intercept_)
print(model.coef_)
```

#### Model Tuning (Model Dogrulama)

```python
#sinama seti
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20, random_state = 99 )
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
#egitim hatasi
np.sqrt(mean_squared_error(y_train, model.predict(X_train)))
```

```python
#test hatasi
np.sqrt(mean_squared_error(y_test, model.predict(X_test)))
```

```python
#k-katli cross validation
from sklearn.model_selection import cross_val_score
cross_val_score(model, X_train, y_train , cv = 10 , scoring = 'neg_mean_squared_error')
```

```python
#k-katli cross validation
from sklearn.model_selection import cross_val_score
cross_val_score(model, X_train, y_train , cv = 20 , scoring = 'neg_mean_squared_error')
```

Her adimda train seti cv sayisina bolunur cv sayisinin 1 eksigiyle model kurulur disarda kalan bir parcayla tahmin edilir. bu sekilde cv kadar hata elde edilir.

```python
#cross validation ile MSE
np.mean(-cross_val_score(model, X_train, y_train , cv = 20 , scoring = 'neg_mean_squared_error'))
```

```python
#cross validation ile RMSE
np.sqrt(np.mean(-cross_val_score(model, X_train, y_train , cv = 20 , scoring = 'neg_mean_squared_error')))
```
