## Introduction



This notebook analyses a data frame that is composed of 1834 rows every row is about a user, and 34 columns. The data frame  contains mainly of 3 things: 1- measurements of 5 personality traits of the users 2- Information about movie lists provided to them 3- users opinion about the movie lists. The personality model that the userâ€™s personality was assessed based upon is called the big five personality scale and it is the most well researched personality model. From its name you know that it is comprised of 5 traits:

1. extraversion/introversion
2. Agreeableness
3. consciounsesness
4. neuroticism/emotional stability
5. openness to experience

For more information visit: https://www.kaggle.com/arslanali4343/top-personality-dataset/discussion/187193

Now, I will try to answer the same question but i will use different aproach. I will create 6 new data frames each one with high/low in openness, extraversion, and conscientiousness for every assigned metric, so in total there will be 18 data frames. The indipindant varible in the coming analysis would be the level of popularity or serendipity or the diversity and the dependant variable would be 'enjoyed watching'.

The way that I will make the new data frames and decide if certain user is high or low in a certain trait would be with using the average and standard deviation. if a certain user is one standard deviation higher than the average in a certain trait than he/she would be included in the high data frames.

```python
# Here the assigned conditions are replaced by numbers 

df['assigned condition'] = df['assigned condition'].replace(['low'], 0.0)
df['assigned condition'] = df['assigned condition'].replace(['medium'], 1.0)
df['assigned condition'] = df['assigned condition'].replace(['high'], 2.0)
```

```python
df = df[df['assigned condition'] != ' default']
```

```python


df_high_openness_diversity = df[(df['openness'] > (df['openness'].mean() + df['openness'].std())) 
                                & (df['assigned metric'] == 'diversity')]
df_high_openness_popularity = df[(df['openness'] > (df['openness'].mean() + df['openness'].std()))
                                 & (df['assigned metric'] == 'popularity')]
df_high_openness_serendipity = df[(df['openness'] > (df['openness'].mean() + df['openness'].std())) 
                                  & (df['assigned metric'] == 'serendipity')]
df_low_openness_diversity = df[(df['openness'] < (df['openness'].mean() - df['openness'].std())) 
                               & (df['assigned metric'] == 'diversity')]
df_low_openness_popularity = df[(df['openness'] < (df['openness'].mean() - df['openness'].std())) 
                                & (df['assigned metric'] == 'popularity')]
df_low_openness_serendipity = df[(df['openness'] < (df['openness'].mean() - df['openness'].std())) 
                                 & (df['assigned metric'] == 'serendipity')]
```

```python


df_high_conscientiousness_diversity = df[(df['conscientiousness'] > (df['conscientiousness'].mean() + df['conscientiousness'].std()))
                                         & (df['assigned metric'] == 'diversity')]
df_high_conscientiousness_popularity = df[(df['conscientiousness'] > (df['conscientiousness'].mean() + df['conscientiousness'].std()))
                                          & (df['assigned metric'] == 'popularity')]
df_high_conscientiousness_serendipity = df[(df['conscientiousness'] > (df['conscientiousness'].mean() + df['conscientiousness'].std()))
                                           & (df['assigned metric'] == 'serendipity')]
df_low_conscientiousness_diversity = df[(df['conscientiousness'] < (df['conscientiousness'].mean() - df['conscientiousness'].std()))
                                        & (df['assigned metric'] == 'diversity')]
df_low_conscientiousness_popularity = df[(df['conscientiousness'] < (df['conscientiousness'].mean() - df['conscientiousness'].std()))
                                         & (df['assigned metric'] == 'popularity')]
df_low_conscientiousness_serendipity = df[(df['conscientiousness'] < (df['conscientiousness'].mean() - df['conscientiousness'].std())) 
                                          & (df['assigned metric'] == 'serendipity')]
```

```python
df_high_extraversion_diversity = df[(df['extraversion'] > (df['extraversion'].mean() + df['extraversion'].std()))
                                    & (df['assigned metric'] == 'diversity')]
df_high_extraversion_popularity = df[(df['extraversion'] > (df['extraversion'].mean() + df['extraversion'].std()))
                                     & (df['assigned metric'] == 'popularity')]
df_high_extraversion_serendipity = df[(df['extraversion'] > (df['extraversion'].mean() + df['extraversion'].std()))
                                      & (df['assigned metric'] == 'serendipity')]
df_low_extraversion_diversity = df[(df['extraversion'] < (df['extraversion'].mean() - df['extraversion'].std())) 
                                   & (df['assigned metric'] == 'diversity')]
df_low_extraversion_popularity = df[(df['extraversion'] < (df['extraversion'].mean() - df['extraversion'].std()))
                                    & (df['assigned metric'] == 'popularity')]
df_low_extraversion_serendipity = df[(df['extraversion'] < (df['extraversion'].mean() - df['extraversion'].std())) 
                                     & (df['assigned metric'] == 'serendipity')]
```

# knowing the user's personality and the type of list that is given to him/her how close  can our prediction of user's satisfaction be? 



```python
features_personality = ['openness', 'agreeableness', 'emotional_stability', 'conscientiousness', 'extraversion', 'assigned condition', 'assigned metric']

x_personality = df[features_personality]
y = df['enjoy_watching']

from sklearn.preprocessing import LabelEncoder

for i in features_personality:
    label_encoder = LabelEncoder()
    df[i] = label_encoder.fit_transform(df[i])
```

```python

x_personality = df[features_personality]
y = df['enjoy_watching']
```

```python
from sklearn.model_selection import train_test_split

xtrain, xvalid, ytrain, yvalid = train_test_split(x_personality, y, random_state=1)
    
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBRegressor
from lightgbm import LGBMClassifier
```

```python
x_personality.info()
```

```python
error_xgb = []
xtrain, xvalid, ytrain, yvalid = train_test_split(x_personality, y, random_state=1)
for i in range(1,1000,5):
    

 my_model =XGBRegressor(n_estimators= i)
 my_model.fit(xtrain, ytrain)
 predict = my_model.predict(xvalid)
 error_xgb.append(mean_absolute_error(predict, yvalid))
    
plt.plot(range(1,1000,5), error_xgb)
```

```python
min(error_xgb)
```
