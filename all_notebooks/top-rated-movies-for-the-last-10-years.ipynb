```python
links_df = pd.read_csv("/kaggle/input/movielens/links.csv")
movies_df = pd.read_csv("/kaggle/input/movielens/movies.csv")
ratings_df = pd.read_csv("/kaggle/input/movielens/ratings.csv")
tags_df = pd.read_csv("/kaggle/input/movielens/tags.csv")
```

```python
# Split year from the title column in the movies_df DataFrame
movies_df["year"] = movies_df["title"].str.extract('.*\((.*)\).*', expand=True)
movies_df.head()
```

```python
# Convert all timestamps to human readable dates
tags_df["datetime"] = pd.to_datetime(tags_df["timestamp"], unit="s")
ratings_df["datetime"] = pd.to_datetime(ratings_df["timestamp"], unit="s")
print(tags_df.head())
print(ratings_df.head())
```

```python
# Merge movies with ratings DataFrame
ave_ratings = ratings_df.groupby("movieId", as_index=False).mean()
merged_df = movies_df.merge(ave_ratings, on="movieId", how="inner")
merged_df = merged_df[["year","title", "rating", "genres"]]
merged_df.head()
```

```python
years_list = ['1898','1900','1902','1903','1909','1910','1912','1913','1914','1915'
,'1916','1917','1918','1919','1920','1921','1922','1923','1924','1925'
,'1926','1927','1928','1929','1930','1931','1932','1933','1934','1935'
,'1936','1937','1938','1939','1940','1941','1942','1943','1944','1945'
,'1946','1947','1948','1949','1950','1951','1952','1953','1954','1955'
,'1956','1957','1958','1959','1960','1961','1962','1963','1964','1965'
,'1966','1967','1968','1969','1970','1971','1972','1973','1974','1975'
,'1975-1979','1976','1977','1978','1979','1980','1981','1982','1983'
,'1984','1985','1986','1987','1988','1989','1990','1991','1992','1993'
,'1994','1995','1996','1997','1998','1999','2000','2001','2002','2003'
,'2004','2005','2006','2007','2008','2009','2010','2011','2012'
,'2013','2014','2015']

merged_df = merged_df.dropna()
merged_df = merged_df.drop(merged_df[merged_df["year"] == "2007-"].index)
merged_df = merged_df.sort_values("year", ascending=False)

merged_df.head()


top_movie_per_year_list = []

for year in years_list:
    mask4 = merged_df["year"].str.contains(year)
    top_movie = merged_df[mask4].sort_values("rating", ascending=False)
    top_movie = top_movie.iloc[0,1]
    top_movie_per_year_list.append(top_movie)

top_movie_per_year_df = pd.DataFrame()
top_movie_per_year_df["year"] = years_list
top_movie_per_year_df["title"] = top_movie_per_year_list

top_movie_per_year_df = top_movie_per_year_df.sort_values("year", ascending=False)

top_movie_per_year_df[:10]
```
