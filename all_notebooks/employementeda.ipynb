```python
t4to10=data[(data["Year"]>=2004) & (data["Year"]<2011)]
for colname in data.columns:
    if colname== "Year":
        continue
    sns.lineplot(x="Year",y=colname,data=t4to10)
    plt.show()
```

2008 crisis has Huge İmpact. İt is clearly visible.

What about Yearly rate and year.

```python
means=[]
for year in data["Year"]:
    specific_year=data[data["Year"]==year]
    tempsum=0
    for colname in data.columns:
        if colname== "Year":
            continue
        tempsum+=float(specific_year[colname])
    means.append(tempsum/12)         
```

```python
data["Means"]=means
```

```python
sns.lineplot(x="Year",y="Means",data=data)
plt.show()
```

It's always changing, not stable. 

What about Summers?

```python
data.columns
```

```python
summer_means=[]
for year in data["Year"]:
    specific_year=data[data["Year"]==year]
    tempsum=0
    for colname in ["Jun","Jul","Aug"]:
        if colname== "Year":
            continue
        tempsum+=float(specific_year[colname])
    summer_means.append(tempsum/3)
data["summermeans"]=summer_means    
```

```python
sns.lineplot(x="Year",y="summermeans",data=data)
plt.show()
```

Tourism and Agriculture not that important.

```python
difeerence=sum(abs(data["Means"]-data["summermeans"]))
print(difeerence)
print("Year wise: ",difeerence/75)
```

Yes Difference is not that big in Summer and all.

What about winters?

```python
winter_means=[]
for year in data["Year"]:
    specific_year=data[data["Year"]==year]
    tempsum=0
    for colname in ["Jan","Feb","Dec"]:
        if colname== "Year":
            continue
        tempsum+=float(specific_year[colname])
    winter_means.append(tempsum/3)
data["wintermeans"]=winter_means
```

```python
sns.lineplot(x="Year",y="wintermeans",data=data)
plt.show()
```

```python
difeerence=sum(abs(data["Means"]-data["wintermeans"]))
print(difeerence)
print("Year wise: ",difeerence/75)
```

Difference is bigger but still small.

```python
difeerence=sum(abs(data["summermeans"]-data["wintermeans"]))
print(difeerence)
print("Year wise: ",difeerence/75)
```

Winter and Summers has more difference but still small.
