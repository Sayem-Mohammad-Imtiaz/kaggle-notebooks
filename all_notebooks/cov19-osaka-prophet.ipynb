```python
import numpy as np
import pandas as pd
from prophet import Prophet

ds = pd.read_csv("../input/covid19-osaka-city-in-japan-data/covid19_osaka_daily_data.csv")
```

# Predict cumulated sum

```python
ds_pred = pd.DataFrame()
ds_pred["ds"]=ds.date
ds_pred["y"]=ds.cumulative_sum_positive_daily

m = Prophet()
m.fit(ds_pred)
```

```python
future = m.make_future_dataframe(periods=365)
forecast = m.predict(future)
fig1 = m.plot(forecast)
```

```python
from prophet.plot import plot_plotly, plot_components_plotly

plot_plotly(m, forecast)
```

# Predict daily new

```python
ds_pred = pd.DataFrame()
ds_pred["ds"]=ds.date
ds_pred["y"]=ds.positive_daily

m = Prophet()
m.fit(ds_pred)
```
