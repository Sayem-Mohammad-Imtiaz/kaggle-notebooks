```python

from sklearn import ensemble
from sklearn import tree
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
file_name = "../input/bank-personal-loan-modelling/Bank_Personal_Loan_Modelling.csv"

df = pd.read_csv(file_name)
```

```python
df.info()
dfdesc = df.describe()
```

```python
dfcorr = df.corr()
sns.heatmap(dfcorr)

```

# age ile experience 0.99

```python
dfex = df[df["Experience"] < 0]
exval = dfex["Experience"].value_counts()
df["Experience"].replace((-3,-2,-1),(0,0,0), inplace=True)
```

```python
y = df["Personal Loan"]

```

```python
x_data = df.drop(["Personal Loan", "ID", "Age"], axis=1)
```

```python

x = (x_data - np.min(x_data))/(np.max(x_data)-np.min(x_data))
```

```python

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=19)

```

```python
models = [DecisionTreeClassifier(random_state=19),
          RandomForestClassifier(random_state=19),
          GradientBoostingClassifier(random_state=19),
          KNeighborsClassifier(),
          SVC()
          ]


for model in models:
    model.fit(x_train, y_train)
    score1 = model.score(x_test, y_test)
    
    print("_" * 25, str(model))
    print(" ", score1)


```
