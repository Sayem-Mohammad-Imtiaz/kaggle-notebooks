## 1. Import all the required libraries

```python
import pandas as pd
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

## 2. Load and examine the data

```python
df = pd.read_csv('/kaggle/input/mines-vs-rocks/sonar.all-data.csv')
df.head()
```

```python
df.shape
```

## 3. Separate the features and the labels

```python
X_inputs = df.iloc[:, :-1]
y_label = df['R'].values 
```

## 4. Encode labels as numbers

```python
labelencoder_Y = LabelEncoder()
y_label = labelencoder_Y.fit_transform(y_label)
y_label = y_label.reshape([207, 1])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_inputs,y_label,test_size=0.2, random_state=0)
```

## 5. Build the multilayer model with Keras

```python
model = Sequential()
model.add(Dense(300,input_dim=60, activation='relu'))
model.add(Dense(200, activation='relu'))
model.add(Dense(100, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

## 6. Set the training parameters

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

## 7. Train the model

```python
model.fit(X_train,y_train,epochs=30,validation_split=0.1)
```

## 8. Evaluate the model

```python
model.evaluate(X_test,y_test)
```
