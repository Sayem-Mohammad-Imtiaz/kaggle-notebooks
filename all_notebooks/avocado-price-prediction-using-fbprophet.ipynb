```python
df=data_avocado[['Date','AveragePrice','region','type']]
df=df[df.region=='SanFrancisco']
df=df[df.type=='conventional']
df
```

```python
df_SF=df.copy()
df_SF.head(5)
df_SF=df_SF.sort_values(by='Date')
df_SF=df_SF[['Date','AveragePrice']]
data=df_SF.rename(columns={'Date':'ds','AveragePrice':'y'})
```

```python
px.line(df_SF, y='AveragePrice', x='Date')
```

```python
train_data_pr = data[:len(df)-12]
test_data_pr = data[len(df)-12:]
```

```python
from fbprophet import Prophet
```

```python
m = Prophet()
m.fit(train_data_pr)
future = m.make_future_dataframe(periods=52,freq='W')
prophet_pred = m.predict(future)
```

```python
prophet_pred.tail()
```

```python
prophet_pred = pd.DataFrame({"Date" : prophet_pred[-12:]['ds'], "Pred" : prophet_pred[-12:]["yhat"]})
```

```python
Pred=prophet_pred['Pred']
Pred.reset_index(drop=True, inplace=True)
```

```python
Actual=test_data_pr[['ds','y']]
Actual.reset_index(drop=True, inplace=True)
```

```python
Actual['Prediction']=Pred
Final_Pred=Actual
Final_Pred.set_index('ds')
```
