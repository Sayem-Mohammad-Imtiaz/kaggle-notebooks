**Total confirmed number is imported from dataset**

```python
temp_list= []
for b in my_data["Confirmed"]:
    temp_list.append(b)
confirmed = temp_list[-1]
print(confirmed)
```

**Susceptible is calculated**

```python
turkey_population= 84000000
susceptible = turkey_population - confirmed
print(susceptible)
```

```python
my_data["Deaths"]
```

**Total deaths is calculated**

```python
temp_list2= []
for a in my_data["Deaths"]:
    temp_list2.append(a)
deaths = temp_list2[-1]
print(deaths)
```

**Recovered coloumn of the dataset is visualized**

```python
my_data["Recovered"]
```

**Total recovered number is calculated**

```python
temp_list3= []
for c in my_data["Recovered"]:
    temp_list3.append(c)
recovered = temp_list3[-1]
print(recovered)
```

**Infected cases are calculated**

```python
infected = confirmed - recovered - deaths
print(infected)
```

**Recovered + Death cases is calculated**

```python
recovered_fatal = recovered + deaths
print(recovered_fatal)
```

**r0 and rho is calculated as opposed to actual values**

```python
turkey_r0, turkey_rho = (2.5, 0.2)
turkey_sigma = turkey_rho / turkey_r0
turkey_initials = (0.99, 0.00026, 0.00185)
display(Markdown(rf"$\rho = {turkey_rho},\ \sigma = {turkey_sigma}$."))
```

**ODEsimulator is used for the values of the graph**

```python
turkey_population = 84000000
turkey_tau = 1440
sir_param_dict = {
    "rho": turkey_rho, "sigma": turkey_sigma
}
sir_simulator = cs.ODESimulator(country="Turkey", province="SIR")
sir_simulator.add(
    model=cs.SIR, step_n=180, population=turkey_population,
    param_dict=sir_param_dict,
    y0_dict={"Susceptible": susceptible, "Infected": infected, "Fatal or Recovered": recovered_fatal}
)
sir_simulator.run()
sir_simulator.non_dim().tail()
```

**Nnon-dimensional sır model is plotted**
1. **x is the number of susceptible cases per total population**
2. **y is the number of currentry infected cases per total population**
3. ** z is fatal or recovered**

```python
cs.line_plot(
    sir_simulator.non_dim().set_index("t"),
    title=r"SIR: $R_0={0}\ (\rho={1}, \sigma={2})$".format(turkey_r0, turkey_rho, turkey_sigma),
    ylabel="",
    h=1
)
```
