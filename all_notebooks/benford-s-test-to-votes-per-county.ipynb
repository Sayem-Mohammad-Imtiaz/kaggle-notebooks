Simple plot of Benford's law

```python
import pandas as pd
```

```python
df_county =  pd.read_csv('president_county_candidate.csv')
df_county.head()
```

```python
dem_votes = df_county[df_county['party']=='DEM']['votes'].values
rep_votes = df_county[df_county['party']=='REP']['votes'].values
```

```python
total_counties = len(dem_votes)
```

```python
first_digit = {}
for i in range(total_counties):
    take_value = dem_votes[i]
    votes = str(take_value)
    value = votes[0]
    if value in first_digit:
        first_digit[value] += 1
    else:
        first_digit[value] = 1
first_digit.values()
```

```python
# delete the zero
del first_digit['0']
```

```python
import matplotlib.pyplot as plt
lists_d = sorted(first_digit.items())
xd, yd = zip(*lists_d)
plt.plot(xd,yd)
plt.show();
```

```python
first_digit = {}
for i in range(total_counties):
    take_value = rep_votes[i]
    votes = str(take_value)
    value = votes[0]
    if value in first_digit:
        first_digit[value] += 1
    else:
        first_digit[value] = 1
first_digit.values()
```

```python
# delete the zero
del first_digit['0']
```

```python
lists_r = sorted(first_digit.items())
xr, yr = zip(*lists_r)
plt.plot(xr,yr)
plt.show();
```

```python
import numpy as np
from numpy import log10

norm = first_digit['1']/log10(2)
benford = []
for i in range(9):
    i=i+1
    benford.append(log10(1+1/(i)))
```

```python
plt.plot(xr,yr/norm,c='g',label='Trump')
plt.plot(xr,yd/norm,c='b',label='Biden')
plt.plot(xr,benford,c='r',label='Benford')
plt.xlabel('First digit')
plt.ylabel('Frequency')
plt.legend()
plt.show();
```
