```python
#writing the function again to check with another city list
def get_nearest_city(city):
  for check_city in pak_cities:
    if check_city in str(city):
      return check_city
  return city
```

```python

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set()
from datetime import datetime
import plotly.express as px
```

```python
df=pd.read_csv('../input/gufhtugu-publications-dataset-challenge/GP Orders - 5.csv')

df.head()
```

```python
df.rename({"Order Number" : "Order_Number" , "Order Status" : "Order_Status"  , "Book Name" : "Book_Name" ,
           "Order Date" : "Order_Date" , "City (Billing)" : "City" } , axis = 1 , inplace = True)
```

```python
df.dropna(axis=0, inplace= True)
```

```python
df.info()
```

```python
#check the null values in the dataset
df.isnull().sum()
```

```python
row , col  = df.shape

print(f"there are {row} rows and {col} columns")
```

```python
#explore the null values to understand.
df["Book_Name"].head(10)
```

```python
df["Book_Name"].tail(10)
```

```python
df["Order Date & Time"] = pd.DatetimeIndex(df["Order Date & Time"])
df['Date'] = [d.date() for d in df['Order Date & Time']]
df['Time'] = [d.time() for d in df['Order Date & Time']]
df['Month_Name'] = [d.month_name() for d in df['Order Date & Time']]
df["Day_Name"] =[d.day_name() for d in df['Order Date & Time']] 
df['Year'] = [d.year for d in df['Order Date & Time']]
df.head()
```
