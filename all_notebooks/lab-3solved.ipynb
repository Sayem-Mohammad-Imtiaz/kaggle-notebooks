Plot the Number of people survived and did not survive. Hint: Plot the counts values of “Survived” column of dataset

```python
total = len(data.Survived)# total lenght of colum servived
ones=data['Survived'].sum() #sum for colum servived as it gives number of people servived as servived =1,unservived=0
zeros=total-ones #finding number remaing which will be unservived
print('servived',ones,'unsurvived',zeros)
```

```python
import matplotlib.pyplot as plt #import matplot lib for plots
y=[zeros,ones] #y give us total value number
x=["unsurvived","unsurvvied"]
plt.bar(x,y) #ploting the bars
```

```python
g = sns.FacetGrid(data, col='Survived')#using facegrid we will do mini plot of colum survived
g.map(plt.hist, 'Pclass', bins=20) #ploting the Pclass in term of Servived
```

```python
g = sns.FacetGrid(data, col='Survived') #ploting Sex in term of Servived
g.map(plt.hist, 'Sex', bins=20)
```

```python
data.groupby('Survived').mean() #mean of all values in term of Survived using group by 
                                
```
