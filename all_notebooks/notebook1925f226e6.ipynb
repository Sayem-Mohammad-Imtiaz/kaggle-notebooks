```python
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv("../input/credit-card-customers/BankChurners.csv")

# take a look at the dataset
df.head()
```

```python
cdf = df[['Attrition_Flag','Customer_Age','Gender','Dependent_count','Education_Level','Income_Category','Card_Category','Months_on_book','Total_Revolving_Bal','Credit_Limit']]
cdf.head(30)
```

```python
plt.scatter(cdf.Customer_Age, cdf.Months_on_book,  color='red')
plt.xlabel("Customer_Age")
plt.ylabel("Months_on_book")
plt.show()
```

```python
msk = np.random.rand(len(df)) < 0.7
train = cdf[msk]
test = cdf[~msk]
```

```python
plt.scatter(cdf.Customer_Age, cdf.Months_on_book,  color='red')
plt.xlabel("Customer_Age")
plt.ylabel("Months_on_book")
plt.show()
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
x = np.asanyarray(train[['Customer_Age','Months_on_book','Total_Revolving_Bal']])
y = np.asanyarray(train[['Credit_Limit']])
regr.fit (x, y)
# The coefficients
print ('Coefficients: ', regr.coef_)
```

```python
y_hat= regr.predict(test[['Customer_Age','Months_on_book','Total_Revolving_Bal']])
regr = linear_model.LinearRegression()
x = np.asanyarray(test[['Customer_Age','Months_on_book','Total_Revolving_Bal']])
y = np.asanyarray(test[['Credit_Limit']])
print("Sum of squares: %.4f"
      % np.mean((y_hat - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.4f' % regr.score(x, y))
```
