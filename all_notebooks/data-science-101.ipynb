1. ***Pokemon data setimizi tanımlıyoruz. Data Setimizin kolonlarını ve örnek verilerini inceliyoruz.***

```python
data_pokemon = pd.read_csv('../input/pokemon-challenge/pokemon.csv')
data_pokemon.head()
```

**Data setimizin Column adlarını ve tiplerini öğreniyoruz**

```python
data_pokemon.columns
```

**Zaten indexli bir iekilde geleceğinden "#" kolonunu siliyorum.**

```python
data_pokemon = data_pokemon.drop("#", axis=1)
data_pokemon.head()
```

```python
data_pokemon.corr()
```

> **# Ne kadar 1'e yakınsa o kadar doğru orantılıdır.**

# #Corelation Map

```python
#Corelation Map
f,ax= plt.subplots(figsize=(10,10))
sns.heatmap(data_pokemon.corr(),annot=True,linewidths=.5,fmt='.1f',ax=ax)
plt.show()
```

```python
data_pokemon.Speed.plot(kind='line',color='g',label='Speed',linewidth=1,alpha=0.5,grid=True,linestyle=':')
data_pokemon.Defense.plot(kind='line',color='r',label='Defense',linewidth=1,alpha=0.5,grid=True,linestyle='-')
plt.show()
```

# #SCATTER PLOT

```python
#SCATTER PLOT
data_pokemon.plot(kind='scatter',x='Attack',y='Defense',alpha=0.5,color='b')
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Attack-Defense Scatter Plot')
plt.show()
```

> **Buradan Attack ile Defens arasında kısmen doğru bir orantı olduğunu söyleyebiliriz.**

# #HISTOGRAM

```python
#HISTOGRAM
data_pokemon.Speed.plot(kind='hist',bins=50,figsize=(10,10),grid=True)
plt.show()
#plt.clf()  clear plt
```
