```python
train = pd.read_csv('/kaggle/input/jane-street-preprocessed-train/Jane Street Preprocessed.csv')
```

```python
train = train.fillna(train.mean())
```

```python
train['action'] = ((train['resp'] > 0) & (train['weight'] > 0)).astype('int')
train['action'].value_counts()
```

```python
train_X = train.drop(labels='action',axis=1)
train_Y= train['action']
col=train_X.columns
```

```python
train_X = train_X.drop(labels='date',axis=1)
train_X = train_X.drop(labels='weight',axis=1)
train_X = train_X.drop(labels='resp_1',axis=1)
train_X = train_X.drop(labels='resp_2',axis=1)
train_X = train_X.drop(labels='resp_3',axis=1)
train_X = train_X.drop(labels='resp_4',axis=1)
train_X = train_X.drop(labels='resp',axis=1)
```

```python
train_X_norm = (train_X - train_X.mean()) / (train_X.max() - train_X.min())
train_norm = pd.concat([train_X_norm, train_Y], axis=1)
```
