# Data Analysis & Visualization

Seeing the states havin most number of encounter

```python
labels = df['state'].value_counts()[:10].index
values = df['state'].value_counts()[:10].values

colors = df['state']

fig = go.Figure(data=[go.Pie(labels=labels, values=values, textinfo="label+percent",
                             insidetextorientation="radial", marker=dict(colors=colors))])
fig.show()    
```

We can see that California, Texax and Florida has the most number of shootings happening out of the top 10 shooting prone states


```python
import dexplot as dxp
```

```python
dxp.count(val='state', data=df, figsize=(15,3), normalize=True)
```

```python
dxp.count(val='state', data=df, figsize=(15,3), split='gender', normalize=True)
```

From the above data one can  infer that male are more prone to shootings than  the female by US Police

```python
labels = df['manner_of_death'].value_counts()[:].index
values = df['manner_of_death'].value_counts()[:].values

colors = df['manner_of_death']

fig = go.Figure(data=[go.Pie(labels=labels, values=values, textinfo="label+percent",
                             insidetextorientation="radial", marker=dict(colors=colors))])
fig.show()    
```

Almost every person was shot very few were tasered too.

Lets see how many of them were male and female 

```python
dxp.count(val='manner_of_death', data=df, figsize=(8,5), split='gender', normalize=True)
```

Lets analysize the age of the victims

```python
dxp.count(val='age', data=df, figsize=(20,5), normalize=True)
```

```python
df['above40'] = ['above40'if i >=40 else 'below40'for i in df.age]

df['generation']="-"
for i in df.age.index:
    if df['age'][i] >=0 and df['age'][i]<40:
        df['generation'][i]='30s'
    elif df['age'][i] >=40 and df['age'][i]<50:
        df['generation'][i]='40s'
    elif df['age'][i] >=50 and df['age'][i]<60:
        df['generation'][i]='50s'
    else:
        df['generation'][i]='60+'
```

```python
sns.countplot(x=df.above40,  palette="Set2")
plt.ylabel('Number of People', fontsize=10)
plt.title('Age of People', color='red', fontsize=15)
```

Youngers are more prone to the shootings particularly the people in the age group of 25-40 and they even accounts for more than half of the shooting casualities casued by the US Police

Lets visualize which race has been more prone to the shooting by US Police
