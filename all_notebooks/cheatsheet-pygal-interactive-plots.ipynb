<font size=+1 color="black"><u>Attributes used in plots:</u></font>

<font size=+1 color="brown"><b>Basic Chart Attributes:</b></font>
* **title** : Title of plot
* **x_title** : Plot x axis title
* **y_title** : Plot y axis title
* **x_labels / y_labels** : Plot x/y axis label
* **show_x_labels / show_y_labels** : Disable X/Y labels if *False*
* **x_label_rotation / y_label_rotation** : Rotation of X/Y lables
* **width / height** : width/height of plot
* **spacing** : space between elements(eg:Bar plot)
* **margin** : external margin
* **show_legend** : Disable legend if *False*
* **legend_at_bottom** : Display legends if *True*
* **legend_at_bottom_columns** : Display legends in columns(number)
* **include_x_axis** : Include x axis line
* **inverse_y_axis** : Inverse yaxis values
* **range** : range of y axis values
* **tooltip_border_radius** : tooltip box curve(number)

<font size=+1 color="brown"><b>Basic Serie Attributes:</b></font>

* **secondary** : display secondary line of plot(only for serie)
* **stroke** : Display line if *True*
* **fill** : Display area of line if *True*
* **show_dots** :remove dots if *False*
* **dots_size** : size of dot
* **value_formatter** : display the formatted value.(ex:dollar,percentage)

<font size=+1 color="brown"><b>Basic Value Attributes:</b></font>

* **label** : tooltip for particular serie of chart
* **color** : tooltip for particular serie of chart

<font size=+1 color="brown"><b>Basic Styles:</b></font>

* **style** : builtin plot style (ex:DarkStyle) For more styles,click here: http://www.pygal.org/en/stable/documentation/builtin_styles.html
* **background** : The color of the image background 
* **plot_background** :The color of the chart area background
* **foreground** : The main foregrond color
* **opacity** : The opacity of chart element
* **opacity_hover** : The opacity of chart element on mouse hover
* **colors** : The serie color list
* **label_font_size** : The label font size
* **title_font_family** : The title font family

For more custom styles,click here : http://www.pygal.org/en/stable/documentation/custom_styles.html

```python
# This is helper function to render plot in html format
base_html = """
<!DOCTYPE html>
<html>
  <head>
  <script type="text/javascript" src="http://kozea.github.com/pygal.js/javascripts/svg.jquery.js"></script>
  <script type="text/javascript" src="https://kozea.github.io/pygal.js/2.0.x/pygal-tooltips.min.js""></script>
  </head>
  <body>
    <figure>
      {rendered_chart}
    </figure>
  </body>
</html>
"""

def pygalplot(chart):
    rendered_chart = chart.render(is_unicode=True)
    plot_html = base_html.format(rendered_chart=rendered_chart)
    display(HTML(plot_html))
```

<a href="#top" class="btn btn-primary btn-sm" role="button" aria-pressed="true" style="color:white" data-toggle="popover" title="go to Colors">Go to TOC</a>

**Purpose** :  Displays quantitative representation of a variable with custom color based on values.

**Question** : How many international football matches were held by each countries ?(Provide seperate color bar for countries that have conducted more than 100 matches) 

```python
wom_football_df=football['country'].value_counts()[:10].to_frame().reset_index().rename(columns={'index':'country','country':'count'})

# Setting color based on values
color_data = []
for v in wom_football_df['count']:
    if v > 100:
        color_data.append({"value": v, "color": "green"})
    elif v < 101:
        color_data.append({"value": v, "color": "red"})
    else:
        color_data.append(v)
```

```python
# NeonStyle is builtin style of plot ,can try other styles - http://www.pygal.org/en/stable/documentation/builtin_styles.html 
from pygal.style import NeonStyle

wom_football = pygal.Bar(title="Most Women International matches Host Countries",x_title="Countries",y_title="Number of Matches",show_legend=False,style=NeonStyle)
wom_football.x_labels = wom_football_df['country']
wom_football.add('',color_data)
pygalplot(wom_football)
```

**Purpose** : Relationship between variables with respective to time

**Question** : How much are the average ratings for IT,Health & Business Sector jobs over past decade?

```python
jobs_10=jobs[jobs['Founded']>2010]

it=jobs_10[jobs_10['Sector']=='Information Technology'].groupby('Founded').agg({'Rating':'mean'}).reset_index()
years=list(it['Founded'])
years_df=pd.DataFrame(years,columns=['year'])
it_df=pd.merge(years_df,it,right_on="Founded",left_on="year",how="left").replace({np.NAN: None})

business=jobs_10[jobs_10['Sector']=='Business Services'].groupby('Founded').agg({'Rating':'mean'}).reset_index()
business_df=pd.merge(years_df,business,right_on="Founded",left_on="year",how="left").replace({np.NAN: None})

health=jobs_10[jobs_10['Sector']=='Health Care'].groupby('Founded').agg({'Rating':'mean'}).reset_index()
health_df=pd.merge(years_df,health,right_on="Founded",left_on="year",how="left").replace({np.NAN: None})
```

```python
scores = pygal.Line(style=CleanStyle,title="Average Ratings for Sectors",x_title="Year",y_title="Sector Average Ratings")
scores.x_labels = it_df['year']
scores.add('IT',it_df['Rating'])
scores.add('Business',business_df['Rating'])
scores.add('Health',health_df['Rating'])

pygalplot(scores)
```

Histogram are special bars that take 3 values for a bar: the ordinate height, the abscissa start and the abscissa end.

**Purpose** : Display distribution of a continuous variable.

**Question** : What is the closing price distribution of Cipla company?

```python
bins = [100, 300,500, 700, 900,1100,1300,1500]
labels = ['100-300','301-500','501-700','701-900','901-1100','1101-1300','1301-1500']
cipla['freq'] = pd.cut(cipla['Close'], bins=bins, labels=labels)
hist_cip=cipla['freq'].value_counts().to_frame().reset_index()
hist_cip[['index','end']] = hist_cip['index'].str.split('-',expand=True)
hist_cip_list=[]
for x,y,z in zip(hist_cip['freq'],hist_cip['index'],hist_cip['end']):
    hist_cip_list.append((int(x),int(y),int(z)))
```

**Purpose** : Displays numerical representation of a variable with percentage for multiple series in a single gauge chart.

**Question** : How much black and white race people were killed in US among the whole race population?

```python
race_df=shoot['race'].value_counts().to_frame().reset_index().rename(columns={'index':'race','race':'count'})
race_df['percent']=round((race_df['count']/race_df['count'].sum())*100)
black_white=race_df[(race_df['race']=='Black')|(race_df['race']=='White')]

race_data = []
color_gauge=['brown','green']
for v,c in zip(black_white['percent'],color_gauge):
        race_data.append({"value": v, "max_size":100,'color':c})
```
