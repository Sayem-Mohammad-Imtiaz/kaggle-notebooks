```python
locations = [1, 2]
heights = [nospam_cifrao_mean, spam_cifrao_mean]
labels = ["Ham", "Spam"]
plt.bar(locations, heights, tick_label=labels)
plt.title('Taxa média de caracter $ por tipo de email')
plt.ylabel('Taxa média de caracter $')
plt.xlabel('Tipo de email')
```

#### 3. Aplicação do classificador e validação
Utilizou-se o método de classificação Naive Bayes para o problema. E em seguida, pela validação cruzada avaliou-se o desempenho do classificador.

Primeiramente, os dados foram separados em dois vetores, um contendo os parâmetros (features) e outro contendo os rótulos (labels).

```python
features_train = treino.drop(columns=['ham'])
target_train = treino['ham']
```

Foi utilizado o classificador Naive Bayes para a distribuição normal de probabilidade. 

```python
gnb = GaussianNB()

gnb.fit(features_train, target_train)
```

Pelo método da validação cruzada obteve-se 10 pontuações do classificador de modo a avaliar seu desempenho. 

```python
scores = cross_val_score(gnb, features_train, target_train, cv=10)

print(scores)
```

Pelos scores obtidos, avaliou-se o classificador como satisfatório.
