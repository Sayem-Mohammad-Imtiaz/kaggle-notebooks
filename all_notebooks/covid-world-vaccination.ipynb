```python
import pandas as pd

df = pd.read_csv("../input/covid-world-vaccination-progress/country_vaccinations.csv")
```

**1) What vaccines are used and in which countries?**

```python
df1= df.groupby('country')['vaccines'].unique()
df1.head()
```

**2) What country is vaccinated more people?**

```python
df2= df.groupby('country')[['daily_vaccinations']].sum()
print(df2['daily_vaccinations'].idxmax())
df2.reset_index(inplace=True)
```

**3) What country is vaccinated a larger percent from its population?**

```python
pop = pd.read_csv("../input/population-by-country-2020/population_by_country_2020.csv")
pop.set_index('Country (or dependency)', inplace=True)
df2.set_index('country', inplace=True)
df2['population'] = pop['Population (2020)']
df2 = df2.rename(columns={'daily_vaccinations': 'vaccinated'})
df2['vaccinated_percent'] = df2['vaccinated']/df2['population']*100
print(df2['vaccinated_percent'].idxmax())
```
