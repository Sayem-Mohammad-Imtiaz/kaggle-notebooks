## PYTHON LIBRARIES USED:
**ANALYSIS** = *NUMPY* & *PANDAS*

**VISUALIZATION** = *MATPLOTLIB* & *SEABORN*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline 
import seaborn as sns
#importing required libraries
```

```python
df = pd.read_csv("../input/pokemon/Pokemon.csv") #getting the desired dataset file
df.head()
```

### The above Data contains varoius attributes of different pokemons , having parameters :
##### 1. NAME
##### 2. TYPES 
##### 3. TOTAL
##### 4. HP
##### 5. ATTACK
##### 6. DEFENSE
##### 7. SPECIAL ATTACK
##### 8. SPECIAL DEFENSE 
##### 9. GENERATION
##### 10. LEGENDARY

## 1. DATA CLEANING AND GETTING INSIGHTS.

```python
df.set_index("Name",inplace = True) #creating "name" column as index
df.drop(["#"], axis = 1, inplace = True)# removing column "#" from dataset
df
```

```python
print(df.isnull().sum()) #checking missing values in each column
df["Type 2"].fillna(df["Type 1"], inplace = True) #filling missing values in "type2" column with values of "type1" column.
print(df.shape) #getting the number of rows & column of matrix 
```

# 2. DATA VISUALIZATION



## 2.1 CATEGORY

### 2.1.1 NUMBER OF POKEMONS IN *TYPE 1* CATEGORY :

```python
print(df["Type 1"].unique()) #getting all the unique values in "type1" column
plt.hist(df["Type 1"],bins = 18, rwidth = 0.9, linewidth = 2,edgecolor = "black", color = "skyblue")
plt.gcf().set_size_inches(15,8) #getting current figure and then setting the size we wanted
plt.xlabel("Pokemons", fontweight = "bold", fontsize = 20)
plt.ylabel("Counts", fontweight = "bold",fontsize = 20)
plt.title("TYPE 1",fontweight = "bold",fontsize = 15)
```

>>> **Water Pokemon are highest in number in TYPE1 category**

### 2.1.2 NUMBER OF POKEMONS IN *TYPE 2* CATEGORY :

```python
print(df["Type 2"].unique())
plt.hist(df["Type 2"],bins = 18, rwidth = 0.9,linewidth = 2,edgecolor = "black", color = "skyblue")
plt.gcf().set_size_inches(15,8)
plt.xlabel("Pokemons", fontweight = "bold", fontsize = 20)
plt.ylabel("Counts", fontweight = "bold",fontsize = 20)
plt.title("TYPE 2",fontweight = "bold",fontsize = 15)
```

>>> **Flying Pokemon are highest in number in TYPE2 category**

# 2.2 SPEED ⚡⚡

### 2.2.1  TOP 10 FASTEST POKEMONS :

```python
df1 = df.sort_values(by = "Speed",ascending = False) #sorting values of "speed" column in ascending order
df1 = df1.drop_duplicates(subset = ["Speed"],keep = "first") #removing all duplicate values of speed.
df1.head(10)
```

### 2.2.2 SPEED OF DIFFERENT POKEMONS ACROSS GENERATIONS:

```python
sns.violinplot(x = df.Generation , y = df.Speed, linewidth = 4)
plt.gcf().set_size_inches(12,8)
plt.xlabel("Generation", fontweight = "bold", fontsize = 20)
plt.ylabel("Speed", fontweight = "bold",fontsize = 20)
```

>>> **Generation 4 Pokemons are Fastest among all.**

# 2.3 HP

### 2.3.1 HP OF POKEMONS ACROSS TYPE1 CATEGORY:

```python
sns.boxplot(x = df["Type 1"], y = df.HP, linewidth = 3.5,whis = 1, palette = "rocket_r")
plt.gcf().set_size_inches(15,8)
plt.xlabel(" Type 1", fontweight = "bold", fontsize = 20)
plt.ylabel("HP", fontweight = "bold",fontsize = 20)
```

>>> **DRAGON Pokemon has highest HP & GHOST Pokemon has lowest in TYPE1**

### 2.3.2 HP OF POKEMONS ACROSS TYPE2 CATEGORY:

```python
sns.boxplot(x = df["Type 2"], y = df.HP, linewidth = 3.5,whis = 1, palette = "crest")
plt.gcf().set_size_inches(15,8)
plt.xlabel("Type 2", fontweight = "bold", fontsize = 20)
plt.ylabel("HP", fontweight = "bold",fontsize = 20)
```

>>> ***ICE Pokemon has highest HP & BUG Pokemon has lowest in TYPE2** 

# 2.4 FIRE POKEMONS💥   vs   WATER POKEMONS 💧

### 2.4.1 ATTACK VS DEFENSE :

```python
Fire = df[(df["Type 1"]=="Fire") | (df["Type 2"]=="Fire") ] #getting dataset of fire pokemon either in "type1" or "type2" column
Water = df[(df["Type 1"]=="Water") | (df["Type 2"]=="Water") ] #water pokemons
sns.scatterplot(x = Water.Attack.head(50), y = Water.Defense.head(50), label = "Water",marker = "o" , s= 70)
sns.scatterplot(x =Fire.Attack.head(50), y = Fire.Defense.head(50), label = "Fire", marker = "*" , s= 250)
plt.legend() #for printing out the label setted up
plt.gcf().set_size_inches(15,8)
plt.xlabel("ATTACK", fontweight = "bold", fontsize = 20)
plt.ylabel("DEFENSE", fontweight = "bold",fontsize = 20)
plt.grid() #getting the grid design.
```

>>> **FIRE pokemon believe in Attacking more rather Defensing & WATER pokemon does vice-versa.**

# 2.5 TOTAL

### 2.5.1 TOP 5 STRONGEST POKEMONS AMONG ALL.

```python
df2 = df.sort_values(by = "Total", ascending = False)
df2 = df2.drop_duplicates(subset = ["Total"])
df2.head(5)
```

### 2.5.2  STRONGEST  & WEAKEST POKEMONS AMONG TYPE1 CATEGORY:

```python
sns.violinplot(y = df.Total , x = df["Type 1"],palette = "magma",linewidth = 3)
plt.gcf().set_size_inches(20,8)
plt.xlabel("TOTAL", fontweight = "bold", fontsize = 20)
plt.ylabel("TYPE1", fontweight = "bold",fontsize = 20)
plt.grid()
```

>>> **DRAGON pokemon is strongest & BUG pokemon is weakest among TYPE1 category.**

### 2.5.3  STRONGEST & WEAKEST POKEMONS AMONG TYPE2 CATEGORY:

```python
sns.violinplot(y = df.Total , x = df["Type 2"],palette = "crest",linewidth = 3)
plt.gcf().set_size_inches(20,8)
plt.xlabel("TOTAL", fontweight = "bold", fontsize = 20)
plt.ylabel("TYPE2", fontweight = "bold",fontsize = 20)
plt.grid()
```

>>> **DRAGON pokemon is strongest & BUG pokemon is weakest among TYPE2 category.**

### 2.5.4  TOTAL OF POKEMONS IN ALL GENERATION ACROSS LEGENDS :

```python
sns.stripplot(x = "Generation" , y = "Total" , data = df , hue = "Legendary",palette = "husl", size = 7)
plt.gcf().set_size_inches(15,8)
plt.ylabel("TOTAL", fontweight = "bold", fontsize = 20)
plt.xlabel("GENERATION", fontweight = "bold",fontsize = 20)
plt.grid()
```

>>> **WE CAN CLEARLY SEE THAT ALL THE STRONGEST POKEMON ACROSS ALL GENERATIONS ARE LEGENDARY.**

```python
df4 = df.pivot_table(index = "Generation" , values=[ "Sp. Atk","Sp. Def","Speed","HP"],aggfunc = np.mean)
df4
#creating the table showing avg values of diff charcterstics acquired by the pokemons across all generation.
```

### 2.6 GENERAL STATS COMPARISON ACROSS ALL GENERATIONS :

```python
plt.plot( np.arange(1,7) ,  df4["Sp. Atk"],"-ro", markerfacecolor='black',markersize = 10, label = "Special Attack")
plt.plot( np.arange(1,7) ,  df4["Sp. Def"],"-bo", markerfacecolor='black',markersize = 10, label = "Special Defense")
plt.plot( np.arange(1,7) ,  df4["Speed"],"-go", markerfacecolor='black',markersize = 10, label = "Speed")
plt.plot( np.arange(1,7) ,  df4["HP"],"-mo", markerfacecolor='black',markersize = 10, label = "HP")

plt.ylabel("AVERAGE VALUES", fontweight = "bold", fontsize = 20)
plt.xlabel("GENERATION", fontweight = "bold",fontsize = 20)

plt.grid()
plt.gcf().set_size_inches(15,8)
plt.legend()
```

>>> **Clearly Generation 4 pokemons are having brilliant stats among all**

### 2.7 DEFENSE VS ATTACK 

```python
sns.jointplot(data = df , x= "Attack", y = "Defense", kind = "hex" )
plt.grid()
plt.gcf().set_size_inches(10,8)
plt.ylim(0,150)
plt.xlim(0,150)

plt.ylabel("DEFENSE", fontweight = "bold", fontsize = 20)
plt.xlabel("ATTACK", fontweight = "bold",fontsize = 20)
```

>>> **OUR POKEMONS IN THE DATASET ARE BIT MORE ATTACKING THAN DOING DEFENSE.**
