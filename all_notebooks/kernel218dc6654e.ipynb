```python
def return_cof(x,y,m,c):
    _c,_m = 0,0
    n = len(a)
    for i in range(1000):
        for j in range(n):
            #print(x[j],y[j])
            _m += (x[j]*(y[j]-m*x[j]+c))
            _c += (y[j]-(m*x[j]+c))  
        _m += -(2/n)*_m
        _c += -(2/n)*_c
        m += 0.001*_m
        c += 0.001*_c
        #print(m,c)
    print(m,c)
    return m,c
```
