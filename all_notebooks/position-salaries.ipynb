# Data Preprocessing

```python
#Our Features and Our Label
X=position_salaries_data.iloc[:,1:-1].values
y=position_salaries_data.iloc[:,2]

```

```python
#First Let's split our data into train and test set

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
#Let's scale our data
from sklearn.preprocessing import StandardScaler
#Instanciate our Scaler 
scaler=StandardScaler()
#Fit and transform our scaler on our training set
X_train_std=scaler.fit_transform(X_train)
#transform our data on test set 
X_test_std=scaler.transform(X_test)

```

# Let's model

```python
#We are going to use support vector regressor:SVR
from sklearn.svm import SVR 
#Let's instansiate our model
regressor=SVR()
#Let's fit our model
regressor.fit(X_train_std,y_train)
#Let's Predict
y_pred=regressor.predict(X_test_std)
```

```python
score=regressor.score(X_test_std,y_test)

print(score)
```
