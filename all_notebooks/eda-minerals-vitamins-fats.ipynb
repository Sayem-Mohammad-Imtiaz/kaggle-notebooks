```python
df = pd.read_csv("../input/food-nutrition-dataset/food.csv")
df.head(10)
```

```python
#we cannot impute values based on mean, median, mode etc, so it's better to drop missing values
#because food is a sensitive area, we should drop missing values

df = df.dropna(axis = 1)
```

```python
df.shape  #new shape has 4303 rows, 48 columns, it's enough for exploratory data analysis
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df.Category.nunique()  #unique values
```

### Lots of duplicate values, we can pork skins are repeated more than once, get rid of duplicate values

```python
#df2 = df.drop_duplicates(subset=['Category'])  we can also use this, but we want all the values grouped as mean

df2 = df.groupby(['Category'], as_index = False).mean()
df2.head()
```

```python
df2.info()
```

```python
df2.describe()
```

```python
df2[['Category','Data.Carbohydrate', 'Data.Protein']].sort_values('Data.Protein', ascending = False)[:5]
```

### Top 3 food category with high protein are Soy Prot Isolate, Gelatins, Seal with protein quantity 85.69gm, 85.6 gm and 82.6gm respectively

```python
Minerals = ['Data.Major Minerals.Calcium', 'Data.Major Minerals.Copper',
       'Data.Major Minerals.Iron', 'Data.Major Minerals.Magnesium',
       'Data.Major Minerals.Phosphorus', 'Data.Major Minerals.Potassium',
       'Data.Major Minerals.Sodium', 'Data.Major Minerals.Zinc']
Vitamins = ['Data.Vitamins.Vitamin A - IU', 'Data.Vitamins.Vitamin A - RAE',
       'Data.Vitamins.Vitamin B12', 'Data.Vitamins.Vitamin B6',
       'Data.Vitamins.Vitamin C', 'Data.Vitamins.Vitamin E',
       'Data.Vitamins.Vitamin K']
```

```python
sns.pairplot(df2[Minerals])
```

```python
sns.heatmap(df2[Minerals].corr(), annot =True)
```

### High correlation between magnesium and copper(0.6)
### High correlation between magnesium and phosphorous
### Very low negative correlation of sodium with zinc, potassium, iron

```python
sns.pairplot(df2[Vitamins])
```

```python
sns.heatmap(df2[Vitamins].corr(), annot =True)
```
