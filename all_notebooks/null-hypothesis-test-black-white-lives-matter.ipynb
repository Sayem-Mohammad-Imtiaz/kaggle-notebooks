```python
df = pd.read_csv("../input/shootingsbyweek/shootingsbyweek.csv")
df_black_nonblack = pd.read_csv("../input/shootingsbyweek/shootings_black_nonblack.csv")
df_white_nonwhite = pd.read_csv("../input/shootingsbyweek/shootings_white_nonwhite.csv")
```

## Black Lives Matter:

Since the data is uploaded as an effort to discourage racial discrimination specially after *black lives matter* gain momentum so I also started with digging for discrimination done with black race.


```python
df_black = df.loc[df['race'] == 'Black']
df_non_black = df.loc[df['race'] != 'Black']
df_non_black.head()
```

```python
sns.catplot(x="race", y="shoots", kind="box", data=df);
```

```python
sns.catplot(x="race", y="shoots", kind="box", data=df_black_nonblack);
```

## Box Plots:

As you can see there is not much difference between shooting among blacks and other races and hence no sign of discrimination, but still lets confirm by doing a t-test and Infering P-Value out of it.

## T-Test

We are doing a null hypothesis test here where:

Null Hypothesis H0 is: there is no discrimination and all is well.

Alternate Hypothesis: there is discrimination.

```python
from scipy import stats
```

```python
black_list = df_black['shoots'].tolist()
non_black_list = df_non_black['shoots'].tolist()
```

```python
t2, p2 = stats.ttest_ind(black_list,non_black_list)
print("p = " + str(p2))
```

### P-Value

Stagering 23% of P Value revealing that we cannot reject Null Hypothesis.

This also does not mean that we should be accepting null hypothesis its just that we cannot reject it.

## White Lives Matter

In the box plot above you can see there is a massive difference between white and non-white. 

And that is making me run the same discrimination process for Whites.

```python
df_white = df.loc[df['race'] == 'White']
df_non_white = df.loc[df['race'] != 'White']
```

```python
sns.catplot(x="race", y="shoots", kind="box", data=df_white_nonwhite);
```

## T-Test:

Null Hypothesis H0 is: there is no discrimination and all is well.

Alternate Hypothesis: there is discrimination.


```python
white_list = df_white['shoots'].tolist()
non_white_list = df_non_white['shoots'].tolist()
```

```python
t_white, p_white = stats.ttest_ind(white_list,non_white_list)
print("p = " + str(p_white))
```

## P-Value:

Here we have to reject the null hypothesis. There is uncomforting high number of white americans being killed in comparison to other races
