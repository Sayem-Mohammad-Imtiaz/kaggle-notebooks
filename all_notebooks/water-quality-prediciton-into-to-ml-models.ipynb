# Sklearn models

https://scikit-learn.org/stable/user_guide.html

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report , confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('../input/water-potability/water_potability.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df["Potability"].unique()
```

```python
df1 = df.dropna()
```

```python
train_x = df1.drop(columns = df.columns[-1])
train_y = df1[df.columns[-1]]
```

```python
x_train ,  x_test ,y_train , y_test = train_test_split(train_x , train_y , test_size = 0.3 , random_state = 4)
```

# Linear model

```python
from sklearn import linear_model

reg = linear_model.RidgeCV(alphas = np.logspace(-6,6 , 13))
reg.fit(train_x , train_y)
```

```python
reg.alpha_
```

```python
preds = reg.predict(x_test)
preds[1:10]
```

# Naives bayes classifier

```python
from sklearn.naive_bayes import GaussianNB , CategoricalNB
```

```python
Basicgnb = GaussianNB()
y_pred = gnb.fit(x_train , y_train).predict(x_test)
```

```python
print(classification_report(y_pred, y_test , digits=2))
```

```python
# generate and plot confusion matrix
array = confusion_matrix(y_pred, y_test)
df_cm = pd.DataFrame(array)
plt.figure(figsize = (5,5))
sns.heatmap(df_cm, annot=True)
```

# Decision tree

https://scikit-learn.org/stable/modules/tree.html#classification

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train , y_train)
preds =  clf.predict(x_test)
```

```python
print(classification_report(preds , y_test))
```

```python
array = confusion_matrix(preds, y_test)
df_cm = pd.DataFrame(array)
plt.figure(figsize = (5,5))
sns.heatmap(df_cm, annot=True)
```

```python
from sklearn.metrics import plot_roc_curve

svs_disp = plot_roc_curve(clf , x_test , y_test , alpha = 0.7)
```

```python
# Save a photo as image
#tree.plot_tree(clf)


# Save tree as a pdf 
import graphviz

class_names = ['Portability0' , 'Portability1']
feature_names = x_train.columns.to_list()
dot_data = tree.export_graphviz(clf , out_file = None , feature_names = feature_names , class_names = class_names,  rounded =True , filled =True , special_characters = True)

graph = graphviz.Source(dot_data)
graph.render("WaterPol")
graph
```
