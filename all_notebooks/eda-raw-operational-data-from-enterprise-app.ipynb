```python
from plotly import tools, subplots
import plotly.offline as py
py.init_notebook_mode(connected = True)

def line_plot_check_nan(df1, df2, x, y, title, width, height):
    
    trace1 = go.Scatter(
        x = df1[x],
        y = df1[y],
        mode='lines',
        name='with_nans',
        marker = dict(
            color = '#1E90FF', 
        ), 
    )
    
    df3 = df2.dropna()
    trace2 = go.Scatter(
        x = df3[x],
        y = df3[y],
        mode='markers',
        name='no_nans',
        marker = dict(
            color = 'red', 
        ), 
    )
    
    layout = go.Layout(
        title = go.layout.Title(
            text = title,
            x = 0.5
        ),
        font = dict(size = 14),
        width = width,
        height = height,
    )
    
    data = [trace1, trace2]
    fig = go.Figure(data = data, layout = layout)
    py.iplot(fig, filename = 'line_plot')
```

```python
line_plot_check_nan(df,df, 'Date', "\\Swap\\used"," Swap used by date", 1400, 600)
```

```python
line_plot_check_nan(df,df,  'Date', "\\Swap\\free"," Swap used by date", 1400, 600)
```
