### Snowball effect 

The snowball effect is the snowball effect. I think that is the clearest definition because it seems like born of the snowball or quadratic function.

```python
posd = mem.loc[(mem.Diff) > 0]
posdH = posd.loc[(posd.Happy_Sad_group)== "H"]
posdS= posd.loc[(posd.Happy_Sad_group)== "S"]

## Sort time 

newind = (posdH.Mem_Score_Before.sort_values()).index
SposdH =  posdH.reindex(newind)

newind = (posdS.Mem_Score_Before.sort_values()).index
SposdS =  posdS.reindex(newind)
```

```python
layer1 = go.Scatter(x=SposdH.Mem_Score_Before,y=SposdH.Diff,mode="lines+markers",name ="Happy",
                    marker=dict(color="chartreuse"))

layer2 = go.Scatter(x=SposdS.Mem_Score_Before,y=SposdS.Diff,mode="lines+markers",name ="Sad",
                    marker=dict(color="skyblue"))

data  = [layer1,layer2]

layout= dict(title="Snowball effect",xaxis=dict(title= " Memory Score Before",ticklen=6,zeroline=False),yaxis=dict(title="Diff",ticklen=6,zeroline=False))

fig = dict(data = data,layout=layout)

iplot(fig)
```
