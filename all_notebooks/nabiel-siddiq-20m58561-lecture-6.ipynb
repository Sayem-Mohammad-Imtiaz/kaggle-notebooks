In this assignment, case of Brazil is selected.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#read datasets, re-index each row
slt_country1= 'Brazil'
slt_country2= 'Brazil'

data1 = pd.read_csv('../input/covid19-global-dataset/worldometer_coronavirus_daily_data.csv',header=0)
data1=data1[data1['country']==slt_country1]

data2 = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv',header=0)
data2=data2[data2['country']==slt_country2]

data1.index = pd.to_datetime(data1['date'])
data2.index = pd.to_datetime(data2['date'])
```

```python
print(data1.columns)
print(data2.columns)
```

```python
print(np.unique(data1['country']))
print(np.unique(data2['country']))
```

# Analysis
From this point, the program will analyze the data loaded as dataframe object.

```python
#overlay plotting of daily new case and vaccination data
fig,ax1=plt.subplots(figsize=(10,5))
ax1.plot(data1.index,data1['daily_new_cases'])
ax2=ax1.twinx()
ax2.plot(data2.index,data2['daily_vaccinations'],'g')
plt.show()
```

# Improve visualization
Use iplot and plotly to improve visualization. 

```python
from plotly.offline import iplot
import plotly.graph_objs as go

daily_nc = go.Scatter(x=data1.index,y=data1['daily_new_cases'],yaxis='y1',name='Daily New Cases')
daily_v = go.Scatter(x=data2.index,y=data2['daily_vaccinations'],yaxis='y2',name='Daily Vaccinations')

layout_obj = go.Layout(title='Covid-19 vs Vaccine',xaxis=dict(title='Date'),yaxis=dict(title='Daily New Cases'),yaxis2=dict(title="Daily Vaccinations",side='right',overlaying='y'))
fig = go.Figure(data=[daily_nc,daily_v],layout=layout_obj)
iplot(fig)

fig.write_html('results.html')
```

Case of Brazil is selected because recently the government seems to move slowly to vaccinate its people despite the the fact that death cases due to Covid-19 has soared in recent time. Brazil is the second country with highest death toll due to Covid-19 with around 2000 daily deaths. Daily new cases in Brazil is also still showing an upward trend. 

Fortunately, it has been followed by increasing trend of vaccination albeit still too low in terms of quantity to provide nationwide effect to the Corona spreading. In fact, up to now, there is only 9.9 people/hundred of population get vaccinated in Brazil, lagging behind Cambodia and one of its neighbouring country Chile. The government could not also increase the vaccination further because it is running out of vaccine stock. This primarily caused by the previous government policy that was reluctant to buy vaccine from certain countries with only basing the decision on perceptions.

Currently, the government of Brazil is trying to produce the vaccine in-house (licensed by CoronaVac, China). However, many news articles mentioned that the 'active' ingredients of the vaccine is getting more limited and must be imported from China.
