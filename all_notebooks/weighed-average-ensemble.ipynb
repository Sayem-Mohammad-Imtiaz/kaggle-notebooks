## In the ensemble:
* train several models on the same dataset 
* assign weight value to each according to its importance 
* use weighted average of all predictions to make the final prediction

```python
data0 = pd.read_csv('../input/30-days-xgboost-with-5-folds-eda-gpu/submission.csv')
data1 = pd.read_csv('../input/30dml-catboost/submission.csv')
data2 = pd.read_csv('../input/30daysmlxgbrandautoeda/xgb_submission_k.csv')
data3 = pd.read_csv('../input/30dml-lightgbm/submission_lgb_5.csv')
data4 = pd.read_csv('../input/30-days-kfold-xgboost/submission_3.csv')
data5 = pd.read_csv('../input/30-days-of-ml-v1/submission.csv')
data6 = pd.read_csv('../input/lightgbm-with-feature-engineering/submission.csv')
data7 = pd.read_csv('../input/30daysmlxgbrandautoeda/xgb_submission_k.csv')
data8 = pd.read_csv('../input/voting-regression/submission.csv')
data9 = pd.read_csv('../input/30-days-ensemble/1.csv')
data10 = pd.read_csv('../input/30-days-ensemble/2.csv')
data11 = pd.read_csv('../input/30-days-ensemble/3.csv')

data = [data0, data1, data2, data3, data4, data5, data6, data7, data8, data9, data10, data11]
data11.head()
```

```python
# Assigning higher weights to the better scores and trying several values, tone of the best seems to be:
#weights = [0, 0, 0, 0.05, 0.25, 0, 0, 0.7] 
weights = [0, 0, 0, 0, 0, 0.2, 0, 0.4, 0, 0.4, 0, 0]
len(weights)
```

```python
data_final = pd.concat([w*x.target for x, w in zip(data, weights)], axis=1).sum(axis=1)
data_final.name='target'
data_final = pd.concat([data[0]['id'], data_final], axis=1)
data_final.to_csv("submission.csv", index=False)
data_final.head()
```
