```python
#import libraries

import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import Lasso, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score, precision_score,roc_curve,auc
from sklearn.metrics import f1_score
from sklearn import metrics

scaler = MinMaxScaler()
```

```python
data = pd.read_csv('../input/heart-disease-cleveland-uci/heart_cleveland_upload.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
y_data = data['condition']
data.drop("condition", axis=1, inplace=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(data, y_data, test_size=0.33, random_state=11)
```

```python
 regression = LogisticRegression(multi_class='multinomial',
                          solver='newton-cg',
                          random_state=1)
```

```python
regression.fit(scaler.fit_transform(x_train),y_train)
```

```python
score = regression.score(scaler.fit_transform(x_train),y_train)
print(score)
```

```python
score = regression.score(scaler.fit_transform(x_test),y_test)
print(score)
```

```python
def perf_measure(y_actual, y_hat):
    TP = 0
    FP = 0
    TN = 0
    FN = 0

    for i in range(len(y_hat)): 
        if y_actual[i]==y_hat[i]==1:
           TP += 1
        if y_hat[i]==1 and y_actual[i]!=y_hat[i]:
           FP += 1
        if y_actual[i]==y_hat[i]==0:
           TN += 1
        if y_hat[i]==0 and y_actual[i]!=y_hat[i]:
           FN += 1

    return TP, FP, TN, FN
```

```python
TP, FP, TN, FN = perf_measure(np.array(regression.predict(scaler.fit_transform(x_test))),np.array(y_test))
```

```python
print(TP, FP, TN, FN)
```

```python
def get_valuation(TP, FP, TN, FN):
    precison = TP/(TP+FP) 
    recall = TP/(TP+FN)
    F1 = (2*precison*recall)/(precison+recall)
    
    return precison, recall, F1
    
```

```python
precison, recall, F1 = get_valuation(TP, FP, TN, FN)
```

```python
print('Custom Pression score %.2f '%(precison))
print('Custom Recall score %.2f '%(recall))
print('Custom f1 score %.2f '% (F1))
```

```python
y_pred = regression.predict(scaler.fit_transform(x_test))
print('Pression score %.2f '%(metrics.precision_score(y_test, y_pred)))
print('Recall score %.2f '%(metrics.recall_score(y_test, y_pred)))
print('f1 score %.2f '% (metrics.f1_score(y_test, y_pred, average='macro') ))
```

```python
fpr, tpr, thresholds = metrics.roc_curve(y_test, regression.predict(scaler.fit_transform(x_test)))
auc_score = metrics.auc(fpr, tpr)
```
