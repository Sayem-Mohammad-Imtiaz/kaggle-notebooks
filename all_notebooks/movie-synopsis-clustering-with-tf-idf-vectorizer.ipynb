# Data Collection

```python
import pandas as pd
import numpy as np

input_file = "/kaggle/input/movie-synopsis-dataset/movie_synopsis.csv"
df = pd.read_csv(input_file)
df.head()
```

```python
df.info()
```

# Remove Missing Value

```python
index_drop = df[df['synopsis'] == "No overview found."].index
df.drop(index_drop, inplace = True)
df.reset_index(drop = True, inplace = True)
```

```python
df.info()
```

# Preprocessing

```python
df['synopsis'] = df['synopsis'].str.lower()
df.head()
```

# Feature Engineering

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words = 'english') # Use filter to remove english stopwords
features = vectorizer.fit_transform(df['synopsis'].values)
features
```

# Model Training

```python
from sklearn.cluster import KMeans

k = 19 # Number of clusters
model = KMeans(k)
model.fit(features)
```

# Data Labeling

```python
df['cluster'] = model.labels_
df.head()
```

```python
df.to_csv("df_movies_labeled.csv")
```

# Clusters Top Terms

```python
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
n_terms = 10

for i in range(k) :
	print("Cluster %d :" % i)
	for j in order_centroids [i, :n_terms] :
		print('   %s' % terms[j])
	print('----------')
```
