```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.metrics import r2_score, explained_variance_score
```

# Analisis exploratorio y carga del dataset

```python
dataset = pd.read_csv('../input/california-housing-prices-data-extra-features/California_Houses.csv')
dataset.info()
dataset.head()
```

Mapa de calor

```python
sns.heatmap(dataset)
```

```python
x = dataset.iloc[:, 1:14].values
y = dataset.iloc[:, 0].values
```

# Preprocesado

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 5)
```

**Escalado de variables**

```python
from sklearn.preprocessing import StandardScaler 
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

**Reducción de variables con PCA**

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 7)
x_train = pca.fit_transform(x_train)
x_test = pca.fit_transform(x_test)
explained_variance = pca.explained_variance_ratio_
explained_variance
```

# Selección de modelos

**Bosques aleatorios**

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 400)
regressor.fit(x_train, y_train)
y_pred = regressor.predict(x_test)
```

**Calculando el R^2 Ajustado y la puntuación de varianza**

```python
#Porcentaje de precisión del modelo
r2 = r2_score(y_test, y_pred)
r2_square_ajusted = 1 - (1-r2) * (len(y)-1) / (len(y) - x_train.shape[1] - 1)
print(r2_square_ajusted, explained_variance_score(y_test, y_pred))
```

**XGBOOST**

```python
from xgboost import XGBRegressor
regressor2 = XGBRegressor()
regressor2.fit(x_train, y_train)
y_pred2 = regressor2.predict(x_test)
```

**Calculando el R^2 Ajustado y la puntuación de varianza**


```python
r2_square = r2_score(y_test, y_pred2)
r2_square_ajusted2 = 1 - (1-r2_square) * (len(y)-1) / (len(y) - x_train.shape[1] - 1)
print (r2_square_ajusted2, explained_variance_score(y_test, y_pred2))
```
