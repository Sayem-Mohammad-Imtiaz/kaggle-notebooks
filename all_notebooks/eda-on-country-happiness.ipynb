```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import matplotlib.pyplot as plt
import os

for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
five = pd.read_csv("/kaggle/input/world-happiness-report/2015.csv")
```

```python
five.head()
```

Let's just work on 2015 data to understand the basics. After that, we will move on to all of them.

Let's see what affects happiness most.

```python
sns.distplot(five['Happiness Score'])
```

Happiness Score distribution looks like a normal distribution. Let's examine it in each region.

```python
plt.figure(figsize=(20, 10))
sns.barplot(data=five.iloc[:20], x="Happiness Score", y="Country")
plt.title("The happiest 20 country")
```

Apperantly happiest countries are in Europe, but why. Let's see each of the region's happiness scores.

```python
regions = five.groupby("Region").mean().reset_index()
regions
```

```python
plt.figure(figsize=(15, 10))
sns.barplot(data=regions, x="Happiness Score", y="Region")
plt.title("Happiness Scores based on Regions")
```

Let's see the countries in top 3 happiest regions

```python
five[five['Region'] == 'Australia and New Zealand'][['Country', 'Happiness Score', 'Region']]
```

```python
five[five['Region'] == 'North America'][['Country', 'Happiness Score', 'Region']]
```

```python
five[five['Region'] == 'Western Europe'][['Country', 'Happiness Score', 'Region']]
```

Since there are 2 countries in Australia and New Zealand, and North America, I believe the most happiest region is Western Europe since the country number is much higher than other regions.

Let's See the countries in **Sub Saharan Africa**, and why they are the least happy region.

```python
five[five['Region'] == 'Sub-Saharan Africa'][['Country', 'Happiness Score', 'Region', 'Economy (GDP per Capita)']]
```

I believe the reason why this countries are the least happy, beacuse the people in there are too much busy with their survival needs. When we see the happiest countries, their GDP is high and they do not much care about their survival need.

Let's see the correlation between GDP and Happiness

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=five, x="Economy (GDP per Capita)", y="Happiness Score")
```

As we can see from the graph, there is a strong correlation between GDP and Happiness Score. 

```python
regions[['Region', 'Happiness Score', 'Economy (GDP per Capita)']].sort_values(by="Happiness Score", ascending=False)
```

Also the happiest countries should also have higher life expectancy.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=five, x="Happiness Score", y="Health (Life Expectancy)")
```

There is also a strong correlation between life expectancy and happiness. It is quite normal, happiest countries are also the richest countries. Therefore, their health facilites are move advanced than the poor countries. Another reason for people to be happy, ***live long!***.

```python
regions.sort_values(by="Happiness Score", ascending=False)
```

As we can see from table, freedom, trust are directly correlated with happiness.

Let's see a correlation map to prove it. 

```python
corr = five.corr()
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(7, 5))
    ax = sns.heatmap(corr,  cmap="YlGnBu", vmax=.3, square=True, annot=True)
```

Freedom and trust is correlated around 0.5. But the most important factor of happiness score, the most correlated one is the GDP. And I believe it is true. Since the survival needs is the most crucial thing in the life.
