```python
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(data.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
plt.show()
```

```python
data.pH.plot(kind = 'line', color = 'purple',label = 'pH',linewidth=1,alpha = 0.5,grid = True,linestyle = ':')
data.alcohol.plot(color = 'blue',label = 'alcohol',linewidth=1, alpha = 0.5,grid = True,linestyle = '-.')
plt.legend(loc='upper right')    
plt.xlabel('x axis')             
plt.ylabel('y axis')
plt.title('Line Plot')            
plt.show()
```

```python
data.plot(kind='scatter', x='pH', y='alcohol',alpha = 0.5,color = 'blue')
plt.xlabel('pH')          
plt.ylabel('alcohol')
plt.title('pH alcohol Scatter Plot') 
```

```python
data.pH.plot(kind = 'hist',bins = 20,figsize = (12,12))
plt.show()
```
