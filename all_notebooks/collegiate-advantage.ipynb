```python
import matplotlib.pyplot as plt
import pandas as pd
import math
```

```python
types = pd.read_csv('../input/college-salaries/salaries-by-college-type.csv')
types = types.drop(['School Name'], axis = 1)
types
```

```python
def to_int(x):
    """
    Passed to pd.apply to turn all str data to numeric
    """
    if type(x) == float:
        pass
    else:
        num = x[1:]
        num = num.replace(',','')
        num = num.replace('.00','')
        return int(num)
```

```python
types['Starting Median Salary'] = types['Starting Median Salary'].apply(to_int)
types['Mid-Career Median Salary'] = types['Mid-Career Median Salary'].apply(to_int)
types['Mid-Career 10th Percentile Salary'] = types['Mid-Career 10th Percentile Salary'].apply(to_int)
types['Mid-Career 25th Percentile Salary'] = types['Mid-Career 25th Percentile Salary'].apply(to_int)
types['Mid-Career 75th Percentile Salary'] = types['Mid-Career 75th Percentile Salary'].apply(to_int)
types['Mid-Career 90th Percentile Salary'] = types['Mid-Career 90th Percentile Salary'].apply(to_int)
```

```python
grouped_types = types.groupby(['School Type']).mean()
```
