```python
import os
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from imageai.Detection import ObjectDetection
```

```python
def show_image(path_image):
    img = mpimg.imread(path_image)
    imgplot = plt.imshow(img)
    plt.show()

def get_images_path():
    all_image_paths = []
    for dirname, _, filenames in os.walk('/kaggle/input'):
        for filename in filenames:
            if(filename[len(filename)-3:] == "JPG"):
                all_image_paths.append(os.path.join(dirname, filename))

    return all_image_paths


print("Number of images: ", len(get_images_path()))
```

```python
execution_path = os.getcwd()

model_path = "../input/imageaiweighs/yolo.h5"

detector = ObjectDetection()
detector.setModelTypeAsYOLOv3()
detector.setModelPath( os.path.join(execution_path , model_path))
detector.loadModel()
```

```python
# example with an image
detections = detector.detectObjectsFromImage(input_image=os.path.join(execution_path , "../input/tacotrashdataset/data/batch_1/000000.jpg"), output_image_path="./imagenew.jpg", minimum_percentage_probability=30)

for eachObject in detections:
    print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"])
    
show_image("./imagenew.jpg")
```

```python
# example with all image TACO

for image_path in get_images_path():
    detections = detector.detectObjectsFromImage(input_image=os.path.join(execution_path , image_path), output_image_path="./imagenew.jpg", minimum_percentage_probability=30)

    for eachObject in detections:
        print("---- ", image_path)
        print(eachObject["name"] , " : ", eachObject["percentage_probability"], " : ", eachObject["box_points"])
```
