```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import calendar
import seaborn as sns
```

```python
data = pd.read_csv('../input/rainfall-in-pakistan/Rainfall_1901_2016_PAK.csv')
data.head()
```

```python
data['Month_Num'] = data['Month'].apply(lambda x: {month: index for index, month in enumerate(calendar.month_name) if month}[x])
data.head()
```

```python
f,ax=plt.subplots(1,1,figsize=(30,10))
title = plt.title('Rainfall in Monsoon Season', fontsize=20)
title.set_position([0.5, 1.05])
data_monsoon = data[(data["Month_Num"] == 7) | (data["Month_Num"] == 8)]
g = sns.boxplot(data_monsoon["Month"] , data["Rainfall - (MM)"],ax=ax)
```

```python
f,ax=plt.subplots(1,1,figsize=(30,10))
title = plt.title('Monsoon season by Year and Month', fontsize=20)
title.set_position([0.5, 1.05])
g = sns.scatterplot(data=data_monsoon, x=" Year", y="Rainfall - (MM)", hue="Month")
```

### Combining Data of both Months for each Year

```python
data_monsoon_combined = data_monsoon.groupby([" Year"]).agg({'Rainfall - (MM)': "sum"}).reset_index()
data_monsoon_combined
```

```python
f,ax=plt.subplots(1,1,figsize=(30,10))
title = plt.title('Monsoon season Linear Regression', fontsize=20)
title.set_position([0.5, 1.05])
g = sns.regplot(data=data_monsoon_combined, x=" Year", y="Rainfall - (MM)")
```

```python
f,ax=plt.subplots(1,1,figsize=(30,10))
title = plt.title('Residuals', fontsize=20)
title.set_position([0.5, 1.05])
g = sns.residplot(data=data_monsoon_combined, x=" Year", y="Rainfall - (MM)")
```

### Inference

Not fitting in linear model, however in both Linear Regression and in Residual Graphswe can see a **Zigzag** pattern, hence exploring more in this direction
