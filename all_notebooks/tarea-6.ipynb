## A. Naive Bayes

Utilizando el dataset de los Pinguinos `data/penguins.csv`

1. Implemente una classificación de la columna `species` utilizando bayes (4 pts)
2. Realice dos gráficos: Uno mostrando los datos etiquetados correctamente (x: bill_length_mm, y:bill_depth_mm, color: species) y otro con las predicciones (x: bill_length_mm, y:bill_depth_mm, color: prediction) (4 pts)
3. Muestre la matriz de confusión con los nombres de las especies (2 pts)


```python
penguins = pd.read_csv('../input/penguins/penguins.csv')
penguins = penguins.dropna()
penguins[['species']] = penguins[['species']].astype('category')
encoder = OrdinalEncoder()
penguins[['species_id']] = encoder.fit_transform(penguins[['species']])

penguins.head(10)
```

```python
X = penguins[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']]
y = penguins[['species_id']].to_numpy().ravel()
```

```python
model = GaussianNB()
model.fit(X, y)
```

```python
list(penguins["species"].unique())
```

```python
model.score(X, y)
```

```python
y_pred = model.predict(X)
```

```python
fig = make_subplots(rows=1, cols=2)

fig.add_trace(go.Scatter(
    x=penguins['bill_length_mm'], 
    y=penguins['bill_depth_mm'],
    mode='markers',
    marker=dict(color=penguins['species_id'])
), 1, 1)

# - . -

fig.add_trace(go.Scatter(
    x=penguins['bill_length_mm'], 
    y=penguins['bill_depth_mm'],
    mode='markers',
    marker=dict(color=y_pred)
), 1, 2)

fig.update_layout(showlegend=False)
    
fig.show()
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y, y_pred)
cm
```

```python
fig = px.imshow(
    cm,
    x=['Adelie', 'Gentoo', 'Chinstrap'],
    y=['Adelie', 'Gentoo', 'Chinstrap']
)
fig.show()
```

```python
titanic = pd.read_csv('../input/titanic/train.csv')
titanic = titanic.drop(['PassengerId', 'Name', 'Ticket'], axis=1)
titanic[['Sex', 'Cabin', 'Embarked']] = titanic[['Sex', 'Cabin', 'Embarked']].astype('category')
titanic = titanic.dropna()
encoder = OrdinalEncoder()
titanic[['Sex', 'Cabin', 'Embarked']] = encoder.fit_transform(titanic[['Sex', 'Cabin', 'Embarked']])
titanic['Age'].fillna(0, inplace=True)
titanic
```

```python
X = titanic.drop(['Survived'], axis=1) 
y = titanic['Survived']
dtmodel = DecisionTreeClassifier(max_depth=4)
dtmodel.fit(X, y)
```

```python
dtmodel.score(X, y)
```

```python
features = X.columns
classes = ['Died', 'Survived']

plt.figure(figsize=(20,20))
tree.plot_tree(dtmodel, filled=True, feature_names=features)
plt.show()
```
