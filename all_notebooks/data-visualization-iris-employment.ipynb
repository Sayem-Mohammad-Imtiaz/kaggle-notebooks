```python
# to import iris.csv file and making a dataframe from it
iris = pd.read_csv('../input/iriscsv/Iris.csv', names = ['sepal_length','sepal_width','petal_length','petal_width', 'Species'])
iris
```

```python
# dropping first row to give it a proper look
iris = iris.drop('Id', axis = 0)
iris
```

```python
iris.shape
```

```python
iris['Species'].unique()
```

```python
#to import matplotlib library
import matplotlib.pyplot as plt
```

```python
#create a figure and axis of our plot
fig, ax = plt.subplots()

# putting sepal_length against sepal_width
ax.scatter(iris['sepal_length'],iris['sepal_width'])

#to create title and axis
ax.set_title('Iris_Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
plt.show()
```

```python
# create color dictionary
colors = {'Iris-setosa':'r', 'Iris-versicolor':'g', 'Iris-virginica':'b'}
# create a figure and axis
fig, ax = plt.subplots()
# plot each data-point
for i in range(len(iris['sepal_length'])):
    ax.scatter(iris['sepal_length'][i], iris['sepal_width'][i],c=colors[iris['Species'][i]])
# set a title and labels
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')
plt.show()
```

```python
# get columns to plot
columns = iris.columns.drop(['Species'])
# create x data
x_data = range(0, iris.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, iris[column], label=column)
# set title and legend
ax.set_title('Iris Dataset')
ax.legend()
plt.show()
```

```python
#create figures and axes
fig, ax = plt.subplots()

#plot histogram
ax.hist(iris['sepal_length'])

#set title and labels
ax.set_title('sepal_length_size')
ax.set_xlabel('sepal_length')
ax.set_ylabel('frequency')
plt.show()
```

```python
#create figure and axis
fig, ax = plt.subplots()

# count the occurance using pandas value_counts function
data = iris['petal_length'].value_counts()

# get x and y data
petal_length = data.index
frequency = data.values

#create bar chart
ax.barh(petal_length, frequency)

#create title and labels
ax.set_title('Petal Length Number')
ax.set_xlabel('petal_length')
ax.set_ylabel('frequency')
plt.show()
```
