#What is the maximum/minimum amount of DC/AC Power generated in a time interval/day?

```python

#for plant 1
#changing datatype
df_p1["DATE_TIME"]=pd.to_datetime(df_p1["DATE_TIME"], format='%d-%m-%Y %H:%M')
```

```python
#splitting date_time column into date column
df_p1['DATE'] = pd.to_datetime(df_p1['DATE_TIME'],format = '%d-%m-%Y %H:%M').dt.date
```

```python
#plant1 min dc_power
a=df_p1.groupby('DATE')["DC_POWER"].min()
print (a)
```

```python
# plant 1 max dcpower
b=df_p1.groupby('DATE')["DC_POWER"].max()
print (b)
```

```python
#for plant2
#max dc power
df_p3["DATE_TIME"]=pd.to_datetime(df_p3["DATE_TIME"], format='%Y-%m-%d %H:%M:%S')  
df_p3['DATE'] = pd.to_datetime(df_p3['DATE_TIME'],format = '%d-%m-%Y %H:%M').dt.date
c=df_p3.groupby("DATE")["DC_POWER"].max()
print(c)
```

```python
#min dc power
d=df_p3.groupby("DATE")["DC_POWER"].min()
print(d)
```

```python
#forplant1 
#MAX AC POWER
e=df_p1.groupby('DATE')["AC_POWER"].max()
print (e)
```

```python
#min ACPOWER 
f=df_p1.groupby('DATE')["AC_POWER"].min()
print (f)
```

```python
#forplant2 
#MAX AC POWER
g=df_p3.groupby('DATE')["AC_POWER"].max()
print (g)
```

```python
#min ACPOWER 
h=df_p3.groupby('DATE')["AC_POWER"].min()
print (h)
```
