**Rename Column name**

```python
data.rename(columns={'per capita income (US$)':'income'},inplace=True)
```

**Train the model using fit Function**

```python
reg=linear_model.LinearRegression()
reg.fit(data[['year']],data.income)
```

**Predict the Income-per-Capita of Canada in 2025**

```python
reg.predict([[2025]])
```

# Model Accuracy Score

```python
reg.score(data[['year']],data.income)
```
