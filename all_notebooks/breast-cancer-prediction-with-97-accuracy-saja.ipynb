# Data Preproceesing

```python
df=df.drop(['Unnamed: 32'],axis=1)
df=df.drop(['id'],axis=1)
```

```python
df.head(5)
```

# Split data X and Y for Train and Test

```python
y=df[['diagnosis']]
```

```python
y
```

```python
x=df.drop(['diagnosis'],axis=1)
```

```python
x
```

# One hor encoding for comvert text data to numeric data

```python
y=pd.get_dummies(y['diagnosis'],drop_first=True)
```

```python
y
```

# Import machine learning library

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y, test_size=0.25,random_state=0)
```

```python
xtest
```

```python
ytest
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.metrics import r2_score
from sklearn.metrics import roc_curve
from sklearn.metrics import average_precision_score, precision_recall_curve
from sklearn.metrics import auc, plot_precision_recall_curve
from sklearn import metrics
```

```python
clf=DecisionTreeClassifier()
```

```python
clf=clf.fit(xtrain,ytrain)
```

```python
pred1=clf.predict(xtest)
```

```python
pred1
```

```python
print("Accuracy:",metrics.accuracy_score(ytest, pred1))
print("Precision:",metrics.precision_score(ytest, pred1))
print("Recall:",metrics.recall_score(ytest, pred1))
```

```python
cnf_matrix = metrics.confusion_matrix(ytest, pred1)
cnf_matrix
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
```

```python
y_pred_proba = clf.predict_proba(xtest)[::,1]
fpr, tpr, _ = metrics.roc_curve(ytest,  y_pred_proba)
auc = metrics.roc_auc_score(ytest, y_pred_proba)
plt.plot(fpr,tpr,label="data 1, auc="+str(auc))
plt.title('Receiver Operating Characteristic Curve(ROC AUC)', y=1.1)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc=4)
plt.show()
```

```python
average_precision = average_precision_score(ytest, pred1)
print(average_precision)
disp = plot_precision_recall_curve(clf, xtest, ytest)
disp.ax_.set_title('Binary class Precision-Recall curve: '
                   'AP={0:0.2f}'.format(average_precision))
```
