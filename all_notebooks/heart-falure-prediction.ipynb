```python
heart_falure = pd.read_csv('/kaggle/input/heart-failure-clinical-data/heart_failure_clinical_records_dataset.csv')
heart_falure = heart_falure.dropna(axis=0)
```

```python
heart_falure
```

```python
X = heart_falure.iloc[:, 0:12]
y = heart_falure.iloc[:, 12]
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier

heart_falure_model = RandomForestClassifier(random_state=1)

heart_falure_model.fit(train_X, train_y)
```

```python
from sklearn.metrics import accuracy_score

r = accuracy_score(heart_falure_model.predict(val_X), val_y, normalize=False)
print(f'Accuracy: {r / len(val_X) * 100}%')
```
