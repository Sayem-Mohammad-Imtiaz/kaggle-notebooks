```python
df=pd.read_csv('../input/spotify-dataset-19212020-160k-tracks/data_by_genres.csv')
df
```

On se débarasse de la colonne genres

```python
df=df.drop('genres',axis=1)
```

```python
df
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import seaborn as sns
from sklearn.preprocessing import normalize
```

```python
kmeans = KMeans(
    init="random",
    n_clusters=2,
    n_init=10,
    max_iter=300,
    random_state=42
)
```

On normalise les données

```python
colonnes=["acousticness","instrumentalness"]
donnees=normalize(df[colonnes], axis=0, norm='max')
```

```python
kmeans.fit(donnees)
```

kmeans.inertia_ nous donne la plus petite valeur SSE

```python
kmeans.inertia_
```

Nous donne l'attribution de chaque entrée dans une liste.

```python
kmeans.labels_
```

Je vais maintenant visualiser le clustering

```python
resultats=np.zeros((donnees.shape[0], 3))
resultats[:,:2]=donnees
for i in range (np.shape(kmeans.labels_)[0]):
    resultats[i,2]=kmeans.labels_[i]
resultats
```

```python
colonnesbis=list.copy(colonnes)
colonnesbis.append('cluster')
```

```python
df_resultats=pd.DataFrame(list(resultats),columns=colonnesbis)
```

```python
sns.lmplot(x='acousticness', y='instrumentalness', data=df_resultats, fit_reg=False, hue='cluster')
```
