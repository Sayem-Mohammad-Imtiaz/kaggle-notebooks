```python
import pandas as pd
import seaborn as sns
import matplotlib.style as style
style.use('ggplot')
import matplotlib.pyplot as plt
%matplotlib inline 

df = pd.read_csv('../input/avocado.csv')
```

Let's pull a sample of the data and see what we are working with. The data has a lot of information, but this analysis will be focusing on Average Price and Type mainly.

```python
df.sample(10)
```

```python
df.isnull().sum()
```

This data set contains both conventional and organic types of avocados. Let's see the price distribution below. 

```python
#Distribution for Average Price for all Avocado types

df['AveragePrice'].plot(kind='hist', bins=30, color='green', figsize=(10,10), grid=True, title='Frequency Distribution for Avocado Prices')
plt.legend()
plt.xlabel('Average Price')
plt.ylabel('Frequency Count')
```

```python
#Conventional type Average price distribution 

Conventional = df[df['type'] == 'conventional']

Conventional['AveragePrice'].plot(kind='hist', bins=30, color='green',figsize=(10,10), grid=True, title='Frquency Distribution for Conventional Avocado Prices')
plt.legend()
plt.xlabel('Average Price Conventional')
plt.ylabel('Frequency Count')
```

For conventional avocado's we see the price distribution is relatively normal. The conventional avocado price hovers around $1.00.  Now let's check out organic. 

```python
Organic = df[df['type']== 'organic']

Organic['AveragePrice'].plot(kind='hist', bins=30, color='green',figsize=(10,10), grid=True, title='Frquency Distribution for Organic Avocado')
plt.legend()
plt.xlabel('Average Price Organic')
plt.ylabel('Frequency Count')
```

As expected, organic avocados are a little more expensive than conventional types of avocados. But by how much? 

```python
Diffprice= Organic['AveragePrice'].mean() - Conventional['AveragePrice'].mean()

Diffprice
```

On average the price difference is roughly .50cents. #avocadopricesaretoohigh

Now let’s plot prices from 2015 - 2018 by regions. Let’s do Organic first.  

```python
#Price broken down by year and region/state (Organic)
mask = df['type']=='organic'

g = sns.factorplot('AveragePrice','region', data= df[mask],hue='year',size=13,aspect=1,join=False,)
```

2017 was an especially bad year for Organic Avocados in terms of price. Look at San Francisco! 

```python
#Price broken down by year and region/state (Conventional)
mask = df['type']=='conventional'
g = sns.factorplot('AveragePrice','region', data= df[mask], hue='year', size=13, aspect=1, join=False,)
```

Conventional type Avocados also suffered from price spikes in 2017. Luckily 2018 Avocado prices have dropped.
