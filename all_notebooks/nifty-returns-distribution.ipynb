```python
import pandas as pd
nifty_data = pd.read_csv('/kaggle/input/indian-stock-index-eod-data1990-onwards/data_1990_2020/index_data/NIFTY_50.csv');
nifty_data.tail()
```

```python
nifty_data['Date'] = pd.to_datetime(nifty_data['Date'])
nifty_data = nifty_data[nifty_data.Date > "1990-12-01"]
nifty_data.tail()
```

```python
nifty_data.loc[nifty_data['High']==0, ['Open','High','Low']] = nifty_data[nifty_data['High'] == 0].Close
nifty_data
```

```python
def get_diffTimeFrame(data, timeframe, offsetVal):
    ohlc_dict = {'Open':'first', 'High':'max', 'Low':'min', 'Close': 'last'}
    result = data.resample(timeframe, origin=offsetVal, on='Date').apply(ohlc_dict)
    result['Returns%'] = (result['Close'] - result['Open'])*100/result['Open'];
    return result
```

```python
def get_runningdiffTimeFrame(data, timeframe, offsetVal):
    ohlc_dict = {'Open':'first', 'High':'max', 'Low':'min', 'Close': 'last'}
    result = data.resample(timeframe, origin=offsetVal, on='Date').apply(ohlc_dict)
    result['Returns%'] = (result['Close'] - result['Open'])*100/result['Open'];
    return result
```

```python
nifty_weekly = get_diffTimeFrame(nifty_data, 'W', 1);
nifty_weekly.tail(12)
```

```python
#nifty weekly range distribution
nifty_weekly['Returns%'].hist(figsize=[12,8],bins=75)
```

```python
nifty_monthly = get_diffTimeFrame(nifty_data, 'M', 1);
nifty_monthly.tail(12)
```

```python
#nifty monthly range distribution
nifty_monthly['Returns%'].hist(figsize=[12,8],bins=95)
```

```python
nifty_yearly = get_diffTimeFrame(nifty_data, '12M', '1990-01-01')
nifty_yearly
```

```python
#Nifty yearly range distribution
nifty_yearly['Returns%'].hist(figsize=[12,8],bins=15)
```
