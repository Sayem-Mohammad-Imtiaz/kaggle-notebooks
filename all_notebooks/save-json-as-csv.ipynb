```python
import json
import pandas as pd
```

Opening the data as a json and reading in as a dict

```python
import os, json
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        if filename.split(".")[-1] == "json":
            path = dirname + "/" + filename
            with open(path, "r") as f:
                data = json.load(f)
```

Creating a DataFrame with all the moments in each series as an attribute

```python
master = pd.DataFrame()
for i in data["Data"]:
    
    df = pd.DataFrame.from_dict(data["Data"][i]["Cards"]).transpose()
    df["Series"] = i
    master = master.append(df)

```

Saving that DataFrame as csv

```python
master.to_csv("series_info_final.csv")
```

Checking to see if all the informmation that we want is saved
