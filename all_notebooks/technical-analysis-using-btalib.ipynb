```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import os
import btalib #https://btalib.backtrader.com/indicator/  Technical Analysis Library

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
df = pd.read_csv('/kaggle/input/reliance-industries-ril-share-price-19962020/Reliance Industries 1996 to 2020.csv', parse_dates=True, index_col='Date')
df
```

```python
df = df.filter(['Close'])
df
```

```python
sma = btalib.sma(df, period=20)
print(sma.df)
```

```python
print(btalib.sma.alias)
print(sma.params.period)
print(sma.p.period)
print(sma.params['period'])
print(list(sma.params))
print(dict(sma.params))
```

```python
sma200 = btalib.sma(df, period=200)
df["200D_SMA"]=sma200.df
df
```

```python
ema50 = btalib.ema(df, period=50)
df["50D_EMA"]=ema50.df
df
```

```python
ax = df['Close'].plot(title="RIL Stock Price with 20 Day Moving average", fontsize=12)
moving_average = df['200D_SMA'] 
moving_average.plot(label='200D_SMA', ax=ax)

# plotting a line plot with it's default size 
print("Plot in it's default size: ") 

plt.show() 
  
# plotting a line plot after changing it's width and height 
f = plt.figure() 
f.set_figwidth(14) 
f.set_figheight(8) 

ax = df['Close'].plot(title="RIL Stock Price with 20 Day Moving average", fontsize=12)
moving_average = df['200D_SMA'] 
moving_average.plot(label='200D_SMA', ax=ax)
  
print("Plot after re-sizing: ") 
plt.show() 
```
