```python
f,ax = plt.subplots(figsize= (14,14))
sns.heatmap(gamelist.corr(), annot= True, linewidths= 5, fmt= ".1f", ax=ax)
plt.show()
```

```python
#Scatter Plot graphic
gamelist.plot(kind = "scatter", x="Platform", y="Global_Sales", alpha= 0,6, color = "blue")
plt.xlabel("Platform")
plt.ylabel("Global_Sales")
plt.title("the effect of platforms on global sales")
```

```python
import matplotlib.pyplot as plt
minGlobal_Sales = gamelist['Global_Sales'].min()
maxGlobal_Sales = gamelist['Global_Sales'].max()
pl = sns.catplot(x = "Platform",y = "Global_Sales", data = gamelist)
pl.set(ylim=(minGlobal_Sales,maxGlobal_Sales))
fig = plt.gcf()
fig.set_size_inches(50, 50)
```
