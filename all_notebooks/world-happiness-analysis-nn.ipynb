## Analysis of World Happiness Report
### Data of the report shows the ladder score for each country with a regional indicator of the countries. The data also shows some aspects for each country.
### This analysis shows the relation of the happiness with these four aspects:
<ul>
<li>Logged GDP per capita</li>
<li>Social support</li>
<li>Healthy life expectancy</li>
<li>Freedom to make life choices</li>
<li>Generosity</li>
<li>Perceptions of corruptiony</li>
</ul>


```python
df=pd.read_csv('/kaggle/input/world-happiness-report-2021/world-happiness-report-2021.csv')
df.head()
#df.describe()
```

```python
df.rename(columns=lambda x: x.strip().lower().replace(" ", "_"), inplace=True)
df.head()
```

## The number of countries in each region

```python
df['regional_indicator'].value_counts()
```

```python
df_country_region = df.groupby('regional_indicator').country_name.count()
df_country_region
#df_country_region.plot()
df_country_region.plot(kind='bar', figsize=(10,10))
plt.title('Number of countries in each region')
plt.xlabel('Region')
plt.ylabel('Number of Countries');
```

## Plotting the ladder score relation with the region

```python
df.plot(y='regional_indicator', x='ladder_score', kind='scatter', figsize=(10,10));
```

## Mean of ladder score of each region

```python
# Use groupby to get the mean ladder_score for each acidity level
df_mean = df.groupby('regional_indicator')['ladder_score'].mean()
print(df_mean)
df_mean.plot(kind='bar', figsize=(10,10));
```

### We conclude that average score is the highest in North America and NCZ "7". And the least average is in South Asia and Sub-Saharan Africa "4.4".

## Plotting the relation between the ladder score and each aspect

```python
df.plot(x="ladder_score", y="social_support", kind="scatter")
df.plot(x="ladder_score", y="healthy_life_expectancy", kind="scatter")
df.plot(x="ladder_score", y="freedom_to_make_life_choices", kind="scatter");
df.plot(x="ladder_score", y="perceptions_of_corruption", kind="scatter");

```

### Social support, healthy life expectency and freedom to make life choice are positively correlated with happiness score. Perceptions of corruption is negatively correlated with happiness.

## Plotting healthy life expectancy with ladder score scatter plot for each region

```python
plot = sns.jointplot(x = 'healthy_life_expectancy', y = 'ladder_score', data = df, hue = 'regional_indicator', height =10)
    
```
