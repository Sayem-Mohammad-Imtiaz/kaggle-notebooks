### Using Word Cloud, find out what words he liked to use the most 

```python
# WordCloud to see what words he used the most...
words = ''
stopwords = set(STOPWORDS)

# iterate throught the csv file 
for val in trump['insult']:
    val = str(val)
    tokens = val.split()
    
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
        
    words += "".join(tokens)+""
    
wordcloud = WordCloud(width = 800, height = 800,
                     background_color = 'white', 
                     stopwords = stopwords, 
                     min_font_size = 10).generate(words)

# plot the WordCloud image
plt.figure(figsize = (8,8), facecolor = None)
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad = 0)

plt.show()
# He likes to say 'fakenews' the most 
```

```python
# How many different comments he posted his comments on Tweet
trump['insult'].nunique()  
#6729 different varieties of comments he posted 
```

```python
trump['insult'].value_counts()
# 1. Fake News  2. Crooked  3. Sleepy Joe  4. failing  5. Crooked Hillary  
```

```python
# Let's make a piechart with the value_counts()
plt.figure(figsize=(8,8))
top_insults = ['Fake News', 'Crooked', 'Sleepy Joe', 'Failing', 'Crooked Hillary']
numbers = [431, 266, 129, 82, 77]
plt.pie(numbers, labels=top_insults, autopct='%d', shadow=True, startangle=90)
plt.title('Top 5 Insults & Numbers (out of 6729) ')
```

```python
# Make a dataframe only focusing on insults he made per a day 
daily_insult = trump.groupby('date')['insult'].count()
daily_insult = pd.DataFrame(daily_insult)

daily_insult.head()
```

```python
#sns.lineplot(data=daily_insult, x='date',y='insult')
```

```python
sns.scatterplot(data=daily_insult, x="date", y="insult")
# Show the plot
plt.show()
```

```python
# Show insults more than 10 times or more
ten_more = daily_insult[daily_insult['insult'] > 10]
ten_more = pd.DataFrame(ten_more)
ten_more
```

```python
# How many insults he posted more than 10 times or more? 
sns.scatterplot(data=ten_more, x="date", y="insult", hue="insult")
```

# Find out his target from the table

```python
target = trump.groupby('target')['insult'].count()
target = pd.DataFrame(target) # make a dataframe by target
target
```

```python
most_target = target.sort_values(['insult', 'target'], ascending=False).groupby('target').head()
most_target = pd.DataFrame(most_target)
most_target = most_target.head(5)
most_target
```

```python
# plot the most 5 targetted items for insults 
plt.figure(figsize=(150,150))
most_target.plot(lw=2, colormap='jet', marker='.', markersize=10, 
         title='Trump Target Media')
```

### Seems like he loves the media the most. He likes the type of attention.
