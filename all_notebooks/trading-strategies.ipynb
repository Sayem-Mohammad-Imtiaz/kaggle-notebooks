In this notebook we will try to apply some trading strategies to tesla stock

first we will read the csv file

Now we will apply trading strategy called Moving Averages.
I this will plot first moving average of the 55 days and then 21 days.

```python
tesla['ma55'] = tesla['Close'].rolling(55).mean()
tesla['ma21'] = tesla['Close'].rolling(21).mean()

tesla['ma55'].loc['2015-01-01':'2015-12-31'].plot(label='MA55')
tesla['ma21'].loc['2015-01-01':'2015-12-31'].plot(label='MA21')
tesla['Close'].loc['2015-01-01':'2015-12-31'].plot(label='Close')
plt.legend()
plt.show()
```

As stock prices are volatile and they fluctuate on daily basis so we can not predict the future movement on th basis of day to day fluctuations.

To overcome this we use average stock prices of the past.It gives better data for analysis.

In above strategy we took average price of the stock for 21 days and then 55 days.

When MA 21 line crosses MA 55 line from down to up then we will buy a stock and MA 21 line crosses MA 55 line from up to down then we will sell the stock.
