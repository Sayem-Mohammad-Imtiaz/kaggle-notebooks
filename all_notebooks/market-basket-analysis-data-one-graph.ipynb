```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (22,11)
plt.rcParams['font.size']=14

f = open("/kaggle/input/datasets-for-appiori/basket_analysis.csv", "r")

labels = []
tarr = ['Apple', 'Bread', 'Butter', 'Cheese', 'Corn', 'Dill', 'Eggs', 'Ice cream', 'Kidney Beans', 'Milk', 'Nutmeg', 'Onion', 'Sugar', 'Unicorn', 'Yogurt', 'chocolate']

for i in range(0, len(tarr)):
    tarr[i] = [tarr[i], 0]

trues = [0]*len(tarr)

lines = f.readlines()

for i in range(1, len(lines)):
    line = lines[i].replace("\n", "").split(",")
    
    for k in range(1, len(line)):
        if line[k] == "True":
            tarr[k-1][1] += 1

tarr.sort(key=lambda xx: xx[1], reverse=True)

for i in range(0, len(tarr)):
    trues[i] = tarr[i][1]
    labels.append(tarr[i][0])

plt.plot(labels, trues, linewidth=1.618, marker="o", color="#13CDCD", label="True")
plt.title("Total number for truthful")
plt.ylabel("Names", fontsize=12)
plt.xlabel("Number of truthful", fontsize=12)
plt.legend(prop={"size": 16})
plt.show()
    
```
