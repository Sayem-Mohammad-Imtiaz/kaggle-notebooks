Prediction Model - Catboost

```python
x = df[['Age','EstimatedSalary']]
y = df['Purchased']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
```

```python
catboost = CatBoostClassifier(eval_metric='AUC')
catboost.fit(x_train,y_train)
```

```python
accuracy_score(y_test, catboost.predict(x_test))
```

```python
catboost_roc_auc = roc_auc_score(y_test, catboost.predict(x_test))
fpr, tpr, thresholds = roc_curve(y_test, catboost.predict_proba(x_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='AUC (area = %0.2f)' % catboost_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive ')
plt.ylabel('True Positive ')
plt.title('ROC')
plt.show()
```
