# Exploring A&E attendace in Engalnd (UK)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv('/kaggle/input/ae-attendances-england/AE_attendances_england_monthly.csv')
df.head()
```

Let's plot the data for England. We first need to sum the data fro all hospitals.

```python
UK = df.loc[:,['date', 'Total attendances' ]].groupby('date').sum().reset_index()
```

```python
fig, ax = plt.subplots(figsize=(15,10)) 

sns.lineplot(UK.date, UK.iloc[:,1])
sns.despine()

ax.xaxis.set_major_locator(plt.MaxNLocator(30))
fig.autofmt_xdate()
```

The plot shows the number of total attendance in the UK. The COVID-19 outbreak can be nicelyl seen at the start of 2020 when people avoided going to hospital at all costs.

Now let's plot a few hospitals:

```python
london_hospitals = [n for n in list(df.Name.dropna().unique()) if 'London' in n]

london = df[df['Name'].isin(london_hospitals)]
```

```python
fig, ax = plt.subplots(figsize=(15,10)) 

sns.lineplot(x='date', y='Total attendances', data=london, hue='Name')
sns.despine()

ax.xaxis.set_major_locator(plt.MaxNLocator(30))
fig.autofmt_xdate()
```

This shows the total cases for the London hospitals. Other things that could be done is to plot the data geospatially or different values of the data such as specific attendances per unit type.
