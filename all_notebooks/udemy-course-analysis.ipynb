```python
df["date"]=pd.to_datetime(df["published_timestamp"])
df = df.drop(["course_id", "url", "published_timestamp"], axis=1).copy()
```

## Làm sạch dữ liệu

```python
df = df.replace(["Business Ffinance"],"Business Finance")
df = df.replace(["Graphic  Design"],"Graphic Design")
df = df.replace(["Musical Instrument"],"Musical Instruments")
```

## Biểu diễn các biến trong data

```python
Price = px.histogram(data_frame=df, x="price",marginal="box",title="Giá tiền các khóa học Udemy")
Price
```

### Có thể thấy các khóa học được phân bổ không đều. Có tổng cộng 310 khóa học miễn phí. Trung bình giá là 66 USD/khóa, trong khi một nửa số khóa học sẽ dưới 45 USD.

```python
Subscriber = px.histogram(data_frame=df,x="num_subscribers",marginal = "box",title = "Số học viên")
Subscriber
```

### Trong khi số học viên trung bình mỗi khóa là 3197, ít nhất 75% số khóa học chưa có đến 3000 học viên. Khóa học có số học viên cao nhất là 268,923 học viên.

```python
Reviewer = px.histogram(data_frame=df,x="num_reviews",marginal = "box",title = "Số lượt đánh giá")
Reviewer
```

```python
Lectures = px.histogram(data_frame=df,x="num_lectures",marginal = "box",title = "Số bài học trong một khóa")
Lectures
```

```python
Duration = px.histogram(data_frame=df,x="content_duration",marginal = "box",title = "Thời lượng khóa học")
Duration
```

```python
Subject = px.histogram(data_frame=df,x="subject",marginal = "box",title = "Số lượng khóa học theo chủ đề")
Subject
```

### Lĩnh vực phổ biến nhất là Web Development, theo sau bởi Business Finance.

```python
df = df.sort_values("date").reset_index(drop=True)
Databytime = px.line(data_frame=df, x="date", y=["num_subscribers","num_reviews","num_lectures"], title='Tổng số học viên, review và số khóa học theo từng năm')
Databytime.show()
```

### Số lượng tương tác của Udemy đang giảm theo thời gian kể từ khi tạo đỉnh năm 2013.
