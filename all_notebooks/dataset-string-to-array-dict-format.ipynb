## When data is stored in a csv some variables lose their format.
I will summerize the solution I found while working with the data

```python
# Load data, show first 5 rows
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
df = pd.read_csv('/kaggle/input/top-ranked-mangas-myanimelist-mal/top500mangaMAL.csv')
df[['English Title', 'Published Dates', 'Genres', 'Author']].head()
```

```python
# Dataframe information -> Dtype (data type) 
df[['English Title','Published Dates', 'Genres', 'Author']].info()
```

# How to change format from string to array or dict :

```python
firstRowGenres = df['Genres'][0]
firstRowGenres
```

```python
import ast # Abstract Syntax Tree
```

For more information about this check: https://docs.python.org/3/library/ast.html#ast.literal_eval

```python
firstRow = ast.literal_eval(firstRowGenres)
firstRow[0]
```

```python
secondRowDates = df['Published Dates'][1]
secondRowDates
```

```python
datesDict  = ast.literal_eval(secondRowDates)
datesDict['from']
```

```python
datesDict['to']
```

```python
dates =df['Published Dates']
dates
```

```python
# Array for all the dictionaries and how to use them
allDates = []
for date in dates:
    allDates.append(ast.literal_eval(date))

allDates[:5]
```

```python
allDates[0]['from']
```

```python
vagabondAuthors = df['Author'][5]
firstRowAuthors = ast.literal_eval(vagabondAuthors)
firstRowAuthors[0]
```

```python
firstRowAuthors[1]
```

As you can see above some times you have  some unwanted white space at the start or finish, a way to take unwanted things fast is:

```python
firstRowAuthors[1].strip()
```
