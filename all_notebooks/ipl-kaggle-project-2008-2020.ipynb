IPL Data Analysis ( 2008 to 2020 ) Basic EDA

For this analysis , the dataset that we are going to use is **Indian Premier League from 2008 to 2020** which is posted on **Kaggle Data**  set .

This analysis is all about finding out the different insights from the dataset and figure out various answers to help the bussiness decisions.

Lets check the below various bussiness objects
* which team had won maximum matches in each session
* Which team had won by maximum wins
* Which team had won by maximum wickets
* Which team had won matches by marinal score
* Has Toss winning helped in winning matches
* Has Home Ground helped in winning matches

```python
#Lets import the basis libraries for handling the data frames and arrays


import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt # for Visualizations
import seaborn as sns # for Visualization

```

Lets import the IPL data set by Matches

```python
IPL = pd.read_csv("../input/my-first-ipl-project/IPL Matches 2008-2020.csv")
```

```python
#Lets do some EDA for understand the data.
IPL.columns

IPL.dtypes
IPL.shape
IPL.head(10)
IPL.describe()
```

```python
IPL.columns
```

```python
#Lets delete the ID column as its no more required
IPL.isna().sum().plot(kind = 'bar' ,color = 'red')
```

We can see the method field having most of the null values , lets delete the method column

```python
IPL = IPL.drop(columns =['method'])
```

Now lets check for other fields

```python
IPL.isna().sum().plot(kind = 'bar' ,color = 'red')
```

Result_margin and city have maximum NaN values .

```python
#Lets fill the NaN values referring to the Venu name as follows
IPL['city'][IPL['venue'] == 'Sharjah Cricket Stadium'] = IPL['city'].fillna('Sharjah')
IPL['city'][IPL['venue'] == 'Dubai International Cricket Stadium'] = IPL['city'].fillna('Dubai')

#And then the result mat=rgin will be filled with 0.
IPL['result_margin']= IPL['result_margin'].fillna(0)

IPL.isna().sum().plot(kind = 'bar' ,color = 'red')
```

```python
#From this , we came to know there are 4 records where  we don't have proper information about the winner ,result ,player of the match . Hence we can remove these 4 records from the database.
```

```python
IPL = IPL[IPL['winner'].isna() == False]
IPL.reset_index(inplace=True)
IPL = IPL.drop(columns =['index'])

IPL.isna().sum().plot(kind = 'bar' ,color = 'red')
```

Now  all the NaN eliminated and we are good to go with our analysis.
Now Lets rename the Team names as per the standard names of their franchesis.
Need to convert the below names i.e 
     #  Deccan Chargers = Sunrisers Hyderabad - Team Hyderabad
     #  Delhi Daredevils = Delhi Capitals
     # Pune Warriors  = Rising Pune Supergiants


```python
len(IPL.team1.unique())
IPL = IPL.replace(['Deccan Chargers'] ,'Sunrisers Hyderabad')
IPL = IPL.replace(['Delhi Daredevils'] ,'Delhi Capitals')
IPL = IPL.replace(['Pune Warriors'] ,'Rising Pune Supergiants')
IPL = IPL.replace(['Rising Pune Supergiant'] ,'Rising Pune Supergiants')
IPL = IPL.replace(['Chennai Super Kings'] ,'Chennai')
IPL = IPL.replace(['Delhi Capitals'] ,'Delhi')
IPL = IPL.replace(['Gujarat Lions'] ,'Gujarat')
IPL = IPL.replace(['Kings XI Punjab'] ,'Punjab')
IPL = IPL.replace(['Kolkata Knight Riders'] ,'Kolkata')
IPL = IPL.replace(['Rajasthan Royals'] ,'Rajasthan')
IPL = IPL.replace(['Royal Challengers Bangalore'] ,'Bangalore')
IPL = IPL.replace(['Rising Pune Supergiants'] ,'Pune')
IPL = IPL.replace(['Sunrisers Hyderabad'] ,'Hyderabad')
IPL = IPL.replace(['Mumbai Indians'] ,'Mumbai')
IPL = IPL.replace(['Kochi Tuskers Kerala'] ,'Kerala')

len(IPL.team1.unique())
len(IPL.team2.unique())
len(IPL.winner.unique())
```

```python
#In the data set we have date column only , lets categorise the data with IPL season using this date column as follows
IPL.date = pd.to_datetime(IPL.date)

#IPL.iloc[IPL['result_margin'].max()]

n_list = []
for i in range(0 , len(IPL['date'])):
    n_year = IPL['date'][i].strftime("%Y")
    n_list.append(n_year) 

IPL['Season'] = "Season-"+ pd.DataFrame(n_list) #Added a new column as Season
#Dropping the first column
IPL.info()

IPL['Season'].unique()
#There are 13 IPL seasons data available for our analysis
```

#Now lets see if the home advantage contributes its winning chances

```python
team_best_city = []
percentage_wins = []
for team in IPL['winner'].unique():
    data = IPL_Cities[IPL['winner']== team]['city'].value_counts(normalize =True)
    team_best_city.append(f'{team} at {list(data.index)[0]}')
    percentage_wins.append(data.max()*100)
```

```python
#formating a bar plot using a list we have created

sns.barplot(x = team_best_city , y = percentage_wins )
plt.title('Percentage of wins at Home' , fontsize = 15)
plt.axhline(np.mean(percentage_wins), label = 'Mean', linestyle = '--')
plt.ylabel('Percentage of wins', fontsize = 15)
plt.xlabel('Teams at Cities by Wins most', fontsize = 15)
plt.yticks( [0,10,20,30,40],['0%','10%','20%','30%','40%'],fontsize = 15,color = '#6d597a')
plt.xticks(rotation = 15)
plt.legend(fontsize =15)
```

#From the above plot , we can make out an inference that ,a team has 41% of more advantage in winning the matches while playing at their home town

#Lets Analise whether toss winning helped in match winning


```python
Toss_Winning = IPL[IPL['toss_winner'] == IPL['winner']]
```

```python
IPL['Toss_vs_Winner'] = ''
IPL['Toss_vs_Winner'][IPL['toss_winner'] == IPL['winner']] = 'Yes'
IPL['Toss_vs_Winner'][IPL['toss_winner'] != IPL['winner']] = 'No'
```

```python
plt.pie(IPL['Toss_vs_Winner'].value_counts() , autopct = '%1.1f%%' , labels = IPL['Toss_vs_Winner'].value_counts().index)
plt.legend(fontsize = 15)
plt.show()
```

From the above plot we can derive an insight that winning the toss gives significant advantage in winning the matches.
Lets check which decision makes more likely to win the matches


```python
#Toss decision in all the matches
plt.pie(IPL['Toss_Decision'].value_counts() , autopct = '%1.1f%%' , labels = IPL['Toss_Decision'].value_counts().index)
plt.legend(fontsize = 10)
plt.title('Toss decision all the teams' , fontsize = 15)
plt.show()
```

From this graph ,it is clear that 60.7 % of the times toss winning team choose to filed first and only 39.3% times choose to bat first 


```python
#Toss decision by winner team
plt.pie(IPL['Toss_Decision'][IPL['toss_winner'] == IPL['winner']].value_counts() , autopct = '%1.1f%%' , labels = IPL['Toss_Decision'].value_counts().index)
plt.title('Toss decision by the match winner teams' , fontsize = 15)
plt.legend(fontsize = 10)
plt.show()
```

It is proved that the decision of choosing fielding first is the best decision

```python
#Toss decision by Top performance team
plt.pie(IPL['Toss_Decision'][IPL['winner'] == 'Mumbai'].value_counts() , autopct = '%1.1f%%' , labels = IPL['Toss_Decision'].value_counts().index)
plt.title('Toss decision by top performing team in all seasons' , fontsize = 15)
plt.legend(fontsize = 10)
plt.show()
```

```python
sns.catplot(x = 'Toss_Decision' , data = IPL[IPL['result'] != 'Tie'] , col = 'winner',kind = 'count' )
plt.title('Teams who won the toss')
plt.show()
```

From the above plot , we got to see that , except chennai city which is Chinnaswamy stadium ,in all the other pitches choosing fielding is a best decision for winning.
