###   LINEPLOT  SAMPLES

```python
data.likes.plot(kind = 'line' , color = 'Green' , label = 'likes' , linewidth = 2 , alpha = 0.5 ,linestyle =':') 
data.dislikes.plot(kind = 'line' ,color = 'red' , label = 'dislikes',linewidth = 1, alpha = 0.5 ,linestyle = '-.')
plt.legend(loc='upper right')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('likes - dislikes videos  Line Plot')
plt.show()
```

```python
df = data.drop(["video_id"],axis=1)
df.plot()
plt.show()
```

### SCATTER PLOT  SAMPLES

```python
plt.scatter(data.likes,data.dislikes,alpha = 0.5,color = 'red') 
plt.xlabel('dislikes')               
plt.ylabel('likes')
plt.title('likes dislikes -  Scatter Plot')            
plt.show()
```
