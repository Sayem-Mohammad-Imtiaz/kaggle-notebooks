```python
import pandas as pd
df = pd.read_csv('../input/up-crime-data-year-2019/MR data - Compiled Data Set.csv')
df.head(5)
```

```python
df.describe()
```

```python
df=df.drop(['Title'],axis=1)
df=df.drop(['Text'],axis=1)
df=df.fillna(0)
```

```python
col=df.columns
df.columns=[i for i in range(0,len(df.columns))]
df.head(20)
```

```python
d={0:0,'Property Disputes':1,'Family Dispute':2,'Petty Quarrels':3,'Money Disputes':4,'Personal Vendetta':5,'Love Affairs':6,'Casteism':7,'Unknown reasons':8}
```

```python
df[1]=df[1].apply(lambda x : d[x])
df.head(20)
```

```python
df[11]=0
df[12]=0
```

```python
for i in range(0,len(df)):
    if ','in str(df.iloc[i,8]):
        cl=df.iloc[i,8].split(',')
        df.iloc[i,8]=int(cl[0])
        df.iloc[i,11]=int(cl[1])
        if len(cl)>2:
            df.iloc[i,12]=int(cl[2])
    elif int(df.iloc[i,8])>100:
        df.iloc[i,11]=int(df.iloc[i,8])%100
        df.iloc[i,8]=int(df.iloc[i,8])//100
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
v1=0
v2=0
v3=0
for i in range(len(df)):
    if df.iloc[i,0]=="Ghaziabad":
        v1=v1+df.iloc[i,1]+df.iloc[i,2]+df.iloc[i,3]+df.iloc[i,4]+df.iloc[i,5]+df.iloc[i,6]+df.iloc[i,7]+df.iloc[i,10]+df.iloc[i,9]
    elif df.iloc[i,0]=="Kanpur":
        v2=v2+df.iloc[i,1]+df.iloc[i,2]+df.iloc[i,3]+df.iloc[i,4]+df.iloc[i,5]+df.iloc[i,6]+df.iloc[i,7]+df.iloc[i,10]+df.iloc[i,9]
    elif df.iloc[i,0]=="Lucknow":
        v3=v3+df.iloc[i,1]+df.iloc[i,2]+df.iloc[i,3]+df.iloc[i,4]+df.iloc[i,5]+df.iloc[i,6]+df.iloc[i,7]+df.iloc[i,10]+df.iloc[i,9]
victim=list((v1,v2,v3))
city=list(("Ghaziabad","Kanpur","Lucknow"))
plt.bar(city,victim)
plt.xlabel("City")
plt.ylabel("Victims")
plt.title("Crime-City wise data")
plt.show()
```
