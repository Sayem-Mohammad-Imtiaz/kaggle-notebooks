```python
# prepare data frames
fm=firstMorning['Item'].value_counts().head(15)
fA=firstAfternoon['Item'].value_counts().head(15)
sMorning=secondMorning['Item'].value_counts().head(15)
# import graph objects as "go"
import plotly.graph_objs as go
# create trace1 
trace1 = go.Bar(
                x = fm.index,
                y = fm.values,
                name = "First Morning",
                marker = dict(color = 'rgba(255, 174, 255, 0.5)',
                             line=dict(color='rgb(0,0,0)',width=1.5)),
                text=fm.index
                )
trace2 = go.Bar(
                x = fA.index,
                y = fA.values,
                name = "First Afternoon",
                marker = dict(color = 'rgba(255, 255, 128, 0.5)',
                             line=dict(color='rgb(0,0,0)',width=1.5)),
                text=fA.index
                )
trace3 = go.Bar(
                x = sMorning.index,
                y = sMorning.values,
                name = "Night",
                marker = dict(color = 'rgba(123, 255, 60, 0.5)',
                             line=dict(color='rgb(0,0,0)',width=1.5)),
                text=sMorning.index
                )
    
data1 = [trace1,trace2,trace3]
layout = go.Layout(barmode = "group")
fig = go.Figure(data = data1, layout = layout)
iplot(fig)
```

```python
# data preparation
data2016 = data.loc[(data['Date']>='2016-10-30')&(data['Date']<'2017-01-01')]
data2017 = data.loc[(data['Date']>='2017-01-01')&(data['Date']<'2017-04-09')]

trace0 = go.Box(
    y=data2016.Transaction,
    name = '2016 datas of transaction',
    marker = dict(color = 'rgb(12, 12, 140)')
)

    
trace1 = go.Box(
        y=data2017.Transaction,
    name = '2017 datas of transaction',
    marker = dict(
        color = 'rgb(12, 128, 128)',
    )
)

data1 = [trace0, trace1]
iplot(data1)
```

```python
# data preparation

trace1 = go.Histogram(
    x=data2016.Item,
    opacity=0.75,#saydamlık
    name = "2016",
    marker=dict(color='rgba(171, 50, 96, 0.6)'))
trace2 = go.Histogram(
    x=data2017.Item,
    opacity=0.75,
    name = "2017",
    marker=dict(color='rgba(12, 50, 196, 0.6)'))

data1 = [trace1, trace2]
layout = go.Layout(barmode='overlay',
                   title=' Items frequency in 2016 and 2017',
                   xaxis=dict(title='Items'),
                   yaxis=dict( title='Count'),
)
fig = go.Figure(data=data1, layout=layout)
iplot(fig)
```
