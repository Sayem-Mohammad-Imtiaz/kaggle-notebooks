Defining Model and finding loss

```python
number_of_input_vars = 2

#Allocating weights and bias
A = np.random.normal(size=(1,number_of_input_vars))
b = np.random.normal(size=1)

#Defining our model
def model(x_input):
    return A.dot(x_input) + b
### Loss function definition ###
# finding loss
def loss(y_predicted, y_target):
    return ((y_predicted - y_target)**2).sum()/y_predicted.shape[1]
dataset = D[[i for i in x_input]]
data_for_input = dataset.values.T
# msg(A,"\n",b,"\n")
y_predicted = model(data_for_input)
current_loss = loss(y_predicted, y_target)
current_loss
```
