```python
x=df['Fuel Consumption City (L/100 km)']
y=df['CO2 Emissions(g/km)']
x,y
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x,y)
lm.coef_,lm.intercept_
```

```python
df1=df['Fuel Consumption City (L/100 km)']<df['Fuel Consumption City (L/100 km)'].quantile(0.99)
```

```python
x=df1['Fuel Consumption City (L/100 km)']
x=np.array(x).reshape(-1,1)
y=df1['CO2 Emissions(g/km)']
y=np.array(y).reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x,y)
y_pred=lm.predict(x)
```

```python
plt.scatter(df1['Fuel Consumption City (L/100 km)'],df1['CO2 Emissions(g/km)'])
plt.scatter(df1['Fuel Consumption City (L/100 km)'],y_pred)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test = train_test_split(x,y,test_size=0.2)
x_train.shape,y_train.shape, x_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lm=LinearRegression()
lm.fit(x_train,y_train)
```

```python
y_train_pred=lm.predict(x_train)
y_test_pred=lm.predict(x_test)
```

```python
plt.scatter(x,y)
plt.scatter(x_train,y_train_pred)
plt.scatter(x_test,y_test_pred,marker='.')
```
