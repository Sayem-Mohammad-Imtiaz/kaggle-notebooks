```python
df_categorical=df1.select_dtypes(include=['object'])
df_categorical.head()
```

```python
from sklearn import preprocessing
```

```python
from sklearn import preprocessing
le=preprocessing.LabelEncoder()
df_categorical=df_categorical.apply(le.fit_transform)
df_categorical.head()
```

```python
# Next, Concatenate df_categorical dataframe with original df (dataframe)

# first, Drop earlier duplicate columns which had categorical values
df2 = df1.drop(df_categorical.columns,axis=1)
df2 = pd.concat([df2,df_categorical],axis=1)
df2.head()
```

```python
X = df2.drop(['workclass'], axis=1)  #  X will hold all features
y = df2['workclass'] # y will hold target/labels

print(X.shape) #dimensions of input data
print(y.shape) #dimensions of output data
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.3, random_state = 0) 
print(X_train.shape)
print(X_test.shape)
```

```python
features = df2.drop(['workclass'], axis=1)  #  X will hold all features
print(features)
target= df2['workclass'] # y will hold target/labels
print(target)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier1=DecisionTreeClassifier(criterion='gini')
classifier1.fit(X_train,y_train)
print(classifier1)
```

```python
classifier2 = DecisionTreeClassifier(criterion='entropy')  
classifier2.fit(X_train, y_train) 
```

```python
y_pred_1 = classifier1.predict(X_test)  
print(y_pred_1)

y_pred_2 = classifier2.predict(X_test)  
print(y_pred_2)
```

```python
from sklearn.metrics import accuracy_score
acc_1=accuracy_score(y_test,y_pred_1)
print('Accuracy for Gini model:{} %'.format(acc_1*100))

acc_2=accuracy_score(y_test,y_pred_2)
print('Accuracy for Entropy model:{} %'.format(acc_2*100))
```

```python
from sklearn.metrics import classification_report, confusion_matrix  
print(confusion_matrix(y_test, y_pred_1))
print(classification_report(y_test, y_pred_1))
```
