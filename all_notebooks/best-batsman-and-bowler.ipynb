```python
delivers = pd.read_csv("/kaggle/input/ipl-complete-dataset-20082020/IPL Ball-by-Ball 2008-2020.csv")
```

```python
delivers = delivers.sort_values(by=["id", "inning", "over", "ball"]).reset_index()
delivers = delivers.drop(columns = ["index"])
delivers
```

```python
temp = [[delivers.iloc[0][0], delivers.iloc[0][1], 1, delivers.iloc[0][4]], [delivers.iloc[0][0], delivers.iloc[0][1], 2, delivers.iloc[0][5]]]
temp
```

```python
i = 1
for i in range(1, len(delivers)):
#rule for first row of match
    if delivers.iloc[i][0] != delivers.iloc[(i-1)][0]:
        temp.append([delivers.iloc[i][0], delivers.iloc[i][1], 1, delivers.iloc[i][4]]) 
        temp.append([delivers.iloc[i][0], delivers.iloc[i][1], 2, delivers.iloc[i][5]])
#rule for first row of new innings
    elif delivers.iloc[i][1] != delivers.iloc[(i-1)][1]:
        temp.append([delivers.iloc[i][0], delivers.iloc[i][1], 1, delivers.iloc[i][4]]) 
        temp.append([delivers.iloc[i][0], delivers.iloc[i][1], 2, delivers.iloc[i][5]])
#rule for other rows
    elif delivers.iloc[i][4] != delivers.iloc[(i-1)][4] and delivers.iloc[i][4] != delivers.iloc[(i-1)][5]:
        temp.append([delivers.iloc[i][0], delivers.iloc[i][1], (temp[-1][2] + 1), delivers.iloc[i][4]])
    elif delivers.iloc[i][5] != delivers.iloc[(i-1)][4] and delivers.iloc[i][5] != delivers.iloc[(i-1)][5]:
        temp.append([delivers.iloc[i][0], delivers.iloc[i][1], (temp[-1][2] + 1), delivers.iloc[i][5]])
```

```python
batsmen_positions = pd.DataFrame(temp, columns = ['id', 'inning', 'position', 'batsman'])
```
