### This kernel will analyze the following:
    - Average Win Shares Over a Career By Draft Pick
    - Average Years Played in the NBA by Draft Pick
    - Ten Highest Career Win Shares Among Players Drafted After 1978

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
draft = pd.read_csv("draft78.csv", squeeze = True)
season = pd.read_csv("season78.csv", squeeze = True)
draft.head(5)
```

```python
draft.dtypes
```

```python
season.head(5)
```

```python
season["Player"] = season["Player"].str.replace("*", "")
```

```python
season.head(5)
```

```python
season.dtypes
```

```python
len(draft)
```

```python
len(season)
```

```python
merged = draft.merge(season, how = "inner", on = "Player")
```

```python
merged.set_index("Player", inplace = True)
```

```python
merged.loc["Michael Jordan"]
```

```python
merged.nunique()
```

```python
merged[merged["Pick"] <= 30]
```

```python
 nba = merged.groupby("Pick")
```

```python
mask = nba["WS"].mean()
```

```python
mask = mask.head(30)
```

```python
mask2 = nba["Yrs"].mean()
```

```python
mask2 = mask2.head(30)
```

```python
mask.plot(kind = "bar", title = "NBA Win Shares By Draft Pick (Post 1978)")
plt.style.use("ggplot")
```

```python
plt.rcParams["figure.figsize"] = (12,6)
```

```python
mask2.plot(kind = "bar", title = "Average Years Played By Draft Pick (Post 1978)")
plt.style.use("ggplot")
plt.rcParams["figure.figsize"] = (12,6)
```

```python
 best_players = merged.sort_values("WS").groupby("Player")
```

```python
mask3 = best_players["WS"].mean()
```

```python
mask4 = mask3.sort_values(ascending = False).head(10)
```

```python
mask4.plot(kind = "bar", title = "Highest Average Career Win Shares (Post 1978)")
plt.style.use("ggplot")
plt.rcParams["figure.figsize"] = (12,6)
```
