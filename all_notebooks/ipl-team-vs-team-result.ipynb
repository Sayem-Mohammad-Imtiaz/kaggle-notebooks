```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
df = pd.read_csv('/kaggle/input/ipl-complete-dataset-20082020/IPL Matches 2008-2020.csv')
df.head()
```

```python
cities = df['city'].value_counts()
fig = px.bar(x=cities.index,
            y=cities,
            text=(cities/len(df['city'])*100),
            title='City Distribution ',
            height=500,
            width=900)

fig.update_layout(uniformtext_minsize=5, uniformtext_mode='show')
fig.update_traces(textposition='outside', texttemplate='%{text:.2s}%',
                 marker=dict(color=cities, colorscale='teal'))
fig.show()
```

**Findings:**

10% of the total matches has been hosted by Mumbai.


```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
```

```python
years = df['year'].value_counts()

fig = px.bar(x=years.index,
            y=years,
            text=(years/len(df['year'])*100),
            height=500,
            width=900)

fig.update_layout(uniformtext_minsize=5, uniformtext_mode='show')
fig.update_traces(textposition='outside', texttemplate='%{text:.3s}%',
                 marker=dict(color=years, colorscale='mygbm'))
fig.show()
```

**Which Team Won The Most Toss?**

```python
tosswinner = df['toss_winner'].value_counts()

fig = px.bar(x=tosswinner.index,
            y=tosswinner,
            text=(tosswinner/len(df['toss_winner'])*100),
            title='Team With The Most Toss Win',
            height=500,
            width=900)

fig.update_layout(uniformtext_minsize=5, uniformtext_mode='show')
fig.update_traces(textposition='outside', texttemplate='%{text:.3s}%',
                 marker=dict(color=tosswinner, colorscale='peach'))
fig.show()
```

The **Mumbai Indians** team has won the toss the most

```python
win = df['winner'].value_counts()

fig = px.bar(x=win.index,
            y=win,
            text=(win/len(df['winner'])*100),
            title='Team With The Most Win',
            height=500,
            width=900)

fig.update_layout(uniformtext_minsize=5, uniformtext_mode='show')
fig.update_traces(textposition='outside', texttemplate='%{text:.3s}%',
                 marker=dict(color=tosswinner, colorscale='ice'))
fig.show()
```

**Mumbai Indians** has won most of the matches

**Toss & Match Which Team Won The Both The Most?**

```python
match_toss_win = df.loc[df['toss_winner']==df['winner'],'winner'].value_counts().reset_index()
match_toss_win = match_toss_win.rename(columns={'index':'Team'})
fig = px.bar(x=match_toss_win['Team'],
            y=match_toss_win['winner'],
            text=(match_toss_win['winner']),
            title='Team With Both Toss & Match Win ',
            height=500,
            width=900)

fig.update_layout(uniformtext_minsize=5, uniformtext_mode='show')
fig.update_traces(textposition='outside', texttemplate='%{text:.2s}%',
                 marker=dict(color=match_toss_win['winner'], colorscale='deep'))
fig.show() 
```

**What Decision Is Chosen The Most After Winning The Toss, Batting Or Fielding**

```python
decision = df['toss_decision'].value_counts()

fig = px.bar(y=decision.index,
            x=decision,
            text=(decision/len(df['toss_decision'])*100),
            title='Toss Dicision',
            height=400,
            width=700,
            orientation='h')

fig.update_layout(uniformtext_minsize=5, uniformtext_mode='show')
fig.update_traces(textposition='outside', texttemplate='%{text:.2s}%',
                 marker=dict(color=decision, colorscale='temps'))
fig.show() 
```

**Match Result**

```python
results = df['result'].value_counts()

fig = px.bar(y=results.index,
            x=results,
            text=(results/len(df['result'])*100),
            title='Match Result',
            height=400,
            width=800,
            orientation='h')

fig.update_layout(uniformtext_minsize=5, uniformtext_mode='show')
fig.update_traces(textposition='outside', texttemplate='%{text:.3s}%',
                 marker=dict(color=results, colorscale='sunset'))
fig.show() 
```

**Which Team Win Margin Was Greater Than 50 Runs The Most?**

```python
margin = df[df['result_margin']>=50]
win_margin = margin['winner'].value_counts()

fig = px.bar(y=win_margin.index,
            x=win_margin,
            text=(win_margin),
            title='Win Margin Greater Than 50 Runs',
            height=400,
            width=800,
            orientation='h')

fig.update_layout(uniformtext_minsize=5, uniformtext_mode='show')
fig.update_traces(textposition='outside',
                 marker=dict(color=win_margin, colorscale='sunset'))
fig.show() 
```
