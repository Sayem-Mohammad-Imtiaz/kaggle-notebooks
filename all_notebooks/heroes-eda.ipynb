```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns
from wordcloud import WordCloud
import warnings


warnings.filterwarnings('ignore')
style.use('seaborn-dark')
```

Little data cleansing: drop all columns if more than 200 values are NA

```python
df = pd.read_csv('../input/superhero-api-dataset/data.csv')
df = df.drop(columns=[column for column in df.columns if (pd.isna(df[column]).sum())>200])
```

Distplot all of the pure numeric data from the dataframe.

```python
float_df = df.select_dtypes(include=['float64'])
sep_columns = float_df.columns
for i in sep_columns:
    plt.figure()
    plt.title(i)
    sns.distplot(df[i])
    
```

```python
plt.figure(figsize=(15,8))
for i in sep_columns:
    plt.title(i)
    sns.kdeplot(df[i], label=i, shade=True)
plt.legend()
plt.xlabel('Points')
```

Investigate how the eye color is distributed

```python
plt.figure(figsize=(15,8))
plt.title('Eye Color')
plt.bar(df.groupby('appearance__eye-color').count().index, df.groupby('appearance__eye-color').count().id )
plt.xticks(rotation=90)
print('')
```

Most heros have blue eyes, brown took the second place and green the third. 

```python
plt.figure(figsize=(15,8))
plt.title('Hair Color')
sns.barplot(df.groupby('appearance__hair-color').count().index, df.groupby('appearance__hair-color').count().id, color='r')
plt.xticks(rotation=90)
print('')
```

Most heros have black hair, followed by blond and brown. The bald ones also figure the top spots.

```python
df['appearance__height__002_c'] = 0
for i in range(len(df['appearance__height__002'])):
    try:
        df['appearance__height__002_c'][i] = (df['appearance__height__002'][i][:(df['appearance__height__002'][i].index('c'))])
    except:
        pass

plt.figure(figsize=(15,8))
plt.xlabel('Height in cm')
plt.title('Heroes Height')
sns.distplot(df[df.appearance__height__002_c>0]['appearance__height__002_c'], color='g')

```

The graph above states an obvious result: heroes are tall. 

Bellow, we also confirm aother commmom knowledge about heroes weight, they need to be fit to fight the evil.  

```python
df['appearance__weight__002_c'] = 0
for i in range(len(df['appearance__weight__002'])):

    try:
        df['appearance__weight__002_c'][i] = (df['appearance__weight__002'][i][:(df['appearance__weight__002'][i].index('k'))])
    except:
        pass
plt.figure(figsize=(15,8))
plt.xlabel('Height in KG')
plt.title('Heroes Weight')
sns.distplot(df[df.appearance__weight__002_c>0]['appearance__weight__002_c'], color='orange')
```

Now let's check out some wordcloud to describe our "wordy" column  variables from the data

```python
def wordcloud_plot(column, title):
    text = " ".join(string for string in df[column])
    wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
    plt.figure(figsize=(12,8))       
    plt.title(title)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python


str_cols = ['biography__alter-egos', 'work__occupation', 'work__base', 'connections__group-affiliation', 'connections__relatives'] 
for col in str_cols:
    wordcloud_plot(col, col)





```
