# **4. United Kingdom vaccination progress analysis**

```python
uk_pivot = uk_vac.pivot(columns='country',values='daily_vaccinations_per_million')
uk_pivot.head()
```

```python
uk_pivot = uk_pivot.dropna()
uk_pivot.head()
```

```python
plt.figure(figsize=(16,6))
sns.lineplot(data=uk_pivot)
plt.ylabel("number of vaccination/million")
plt.xlabel("Date")
plt.title("UK's Daily Vaccination/million Rate")
plt.show()
```

```python
#North Ireland daily new cases & admissions 
NI = pd.read_csv('/kaggle/input/northirlandnewcasesadmission/Northern Ireland_2021-01-27.csv',index_col='date',parse_dates=True)
NI = NI.drop(['areaType','areaCode','areaName'],axis=1)
NI.info()
```

```python
NI['2021-01-20':'2020-12-01'].newAdmissions.plot()
plt.title('North Ireland Daily Number of new addmission')
plt.ylabel('Number of New Admissions')
plt.show()
```

Admmission rate continued to increase, suspected due to virus varient.  Vaccination impact yet to be seen in data.

```python
#Merging Northern Ireland(NI) vaccination data with new daily cases and admissions 
#step 1 - slice NI vaccination data:
NI_vac=uk_vac.loc[uk_vac.country =='Northern Ireland',:]
NI_vac.head()
```

```python
#step 2 - join on date index
NI_full =NI_vac.join(NI)

NI_full.head()
```

```python

vaccnations = NI_full['daily_vaccinations']
cases = NI_full['newCasesByPublishDate']
admissiosn = NI_full['newAdmissions']


fig,ax1 = plt.subplots(figsize =(10,6))

ax2=ax1.twinx() # 2 y axis due to diffrent scale

ax1.plot(vaccnations, label ='Daily vaccinations',color='g',linestyle='solid')
ax1.plot( cases, label = 'Daily New cases', color ='r' ,linestyle='solid')
ax2.plot(admissiosn, label = 'Daily Admissions', color ='b' ,linestyle='solid')

ax1.set_xlabel('Date')
ax1.set_ylabel('Number of daily Cases/Vaccinations')
ax2.set_ylabel('Number of daily Admissions')

plt.tight_layout()
plt.title('Northen Ireland')
plt.legend()
plt.plot()

plt.show()
```
