# **Face Shield or Eyeglasses Detection**

# **INTRODUCTION**

Airborne contamination is common when a person talks, sneezes, or coughs is transmitted by COVID, the COVID19 coronavirus. There has been heightened debate in recent weeks on the possible usage of face shields or eyeglasses to aid in exiting lockout and preventing more surges in Covid-19 infections.

Face shields are visors made of plastic that cover the whole face. While this form of face covering has the potential to be critical in combating the infection, there are many myths on how they function.

Face shields give the wearer an additional layer of security. Due to the fact that they shield the whole face – lips, nose, and eyes – a large proportion of viral particles are blocked from touching the wearer. Face shields are arguably the most effective method of defending against coughs and sneezes.

Although face shield help liberate out against transmission of the coronavirus, experts advise considering your eyes as well. Eyeglasses could also be a protection to be used aside from using face shields. Eyeglasses can be a substitute to protect the eyes from polluted respiratory droplets.

Shields may thus be very helpful for anyone who come into near conjunction with others on a daily basis – for example, in medical environments. They are also extremely vulnerable in careers that rely heavily on nonverbal communication via facial expressions, such as teaching. They can be beneficial in providing teachers with security and trust, allowing them to return to the classroom atmosphere.
