```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn import svm
```

```python
dataset = pd.read_csv("/kaggle/input/breast-cancer-wisconsin-data/data.csv")

diagnosis = dataset["diagnosis"]
# Substituindo M por 1 e B por 0 
diagnosis = np.where(diagnosis == 'M',1,0)
# Removendo variaveis
dataset.drop(["id","diagnosis","Unnamed: 32"],axis = 1,inplace = True)

# Dividindo a base em treino e teste
x_train, x_test, y_train, y_test = train_test_split(
    dataset, diagnosis, test_size=0.20, random_state=9)

# Treinando a base de treino
svm_classifaier = svm.SVC(kernel="linear")
svm_classifaier.fit(x_train, y_train)
y_pred = svm_classifaier.predict(x_test)

# Confuion Matrix treino
cm = metrics.confusion_matrix(y_test, y_pred)
print(cm)

# Precision treino
print(metrics.classification_report(y_test, y_pred))

#####Testando com a base de dados completa
y_pred = svm_classifaier.predict(dataset)

# Confuion Matrix completa
cm = metrics.confusion_matrix(diagnosis, y_pred)
print(cm)

# Precision treino
print(metrics.classification_report(diagnosis, y_pred))
```

Dados normalizados para a realização de um novo treino no conjunto de dados para verificação de uma possivel melhora nos resultados. 

```python
dataset = pd.read_csv("/kaggle/input/breast-cancer-wisconsin-data/data.csv")

diagnosis = dataset["diagnosis"]
# Substituindo M por 1 e B por 0 
diagnosis = np.where(diagnosis == 'M',1,0)
# Removendo variaveis
dataset.drop(["id","diagnosis","Unnamed: 32"],axis = 1,inplace = True)


##### Normalizando os dados
norm = StandardScaler()
dataset = norm.fit_transform(dataset)

# Dividindo a base em treino e teste
x_train, x_test, y_train, y_test = train_test_split(
    dataset, diagnosis, test_size=0.20, random_state=9)

# Treinando a base
svm_classifaier = svm.SVC(kernel="linear")
svm_classifaier.fit(x_train, y_train)
y_pred = svm_classifaier.predict(x_test)

# Confusion Matrix
cm = metrics.confusion_matrix(y_test, y_pred)
print(cm)

# Precision
print(metrics.classification_report(y_test, y_pred))

#####Testando com a base de dados completa
y_pred = svm_classifaier.predict(dataset)

# Confuion Matrix completa
cm = metrics.confusion_matrix(diagnosis, y_pred)
print(cm)

# Precision treino
print(metrics.classification_report(diagnosis, y_pred))
```

Verificamos uma melhora na matrix de confusão após a normalização dos dados.
