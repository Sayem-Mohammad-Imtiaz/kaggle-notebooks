```python
import matplotlib.pyplot as plt
plt.scatter(data["Number of Units"],data["Manufacturing Cost"])
```

The line fitting the curve cannot be a straight line as it appears to be curved.
We have to apply some transformation on the data

```python
plt.scatter(data["Number of Units"],1/data["Manufacturing Cost"])
```

now we can fit a straight line

```python
data["Manufacturing Cost_inv"] = 1/data["Manufacturing Cost"]
```

```python
x = np.array(data["Number of Units"])
x=x.reshape(-1, 1)
y = data["Manufacturing Cost_inv"] 
x.dtype
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
r_sq = model.score(x, y)
r_sq
```

```python
model.coef_
```
