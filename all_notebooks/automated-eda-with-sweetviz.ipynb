# SweetViz: Automated EDA

**Exploratory data analysis (EDA) is an approach to analyzing data sets to summarize their main characteristics, often with visual methods. A statistical model can be used or not, but primarily EDA is for seeing what the data can tell us beyond the formal modeling or hypothesis testing task.EDA generally takes a lot of time.**

**But,what if we could do it in just a single line of code in just 1 minute.Isn't it amaizing guys!!!Yes!! We could do it using Some librarys of Python.SweetViz is one of them.It's fast!! And it gives EDA of the whole dataset within a few seconds. Sweetviz is an open source Python library that generates beautiful, high-density visualizations to kickstart EDA (Exploratory Data Analysis) with a single line of code. Output is a fully self-contained HTML application.The system is built around quickly visualizing target values and comparing datasets. Its goal is to help quick analysis of target characteristics, training vs testing data, and other such data characterization tasks.**

There are 3 main functions for creating reports in SweetViz:</br>

1.analyze(...)

2.compare(...)

3.compare_intra(...)

# 1)Analyze

To analyze a single dataframe, simply use the analyze(...) function, then the show_html(...) function:

```python
import sweetviz as sv
hotelsfeedback_report = sv.analyze(df)
hotelsfeedback_report.show_html('Analyst.html')
```

By Running the following line, is gonna take you to a new html page,which has been created & it'll gonna save automatically in the following folder,in which you are running the Code.

Just to show you guys, how the HTML file look like i'm gonna import the HTML file here.Let's see how it looks like

```python
from IPython.display import IFrame
IFrame(src = 'Analyst.html',width=1000,height=600)
```

# 2)Compare
Sweetviz can also be used to visualize the comparison of test and train data.Now i'm gonna Comparing two dataframes (e.g. Test vs Training sets).So for Comparing i'm gonna use the Iris Flower Dat

```python
iris = pd.read_csv('../input/iris/Iris.csv')
iris.head(2)
```

```python
#Comparing 1st 200 data
train = iris[:70]
test = iris[70:]
comparison_report = sv.compare(train,test)
comparison_report.show_html('Comparison.html')
```

By Running the following line, is gonna take you to a new html page,which has been created & it'll gonna save automatically in the following folder,in which you are running the Code.

```python
IFrame(src = 'Comparison.html',width=1000,height=500)
```

# 3)compare_intra(...)

Now if you want to compare the analysis between two subpopulations then you can use compare_intra() function from the sweetviz library. Support for this is built in through the compare_intra() function. This function takes a boolean series as one of the arguments, as well as an explicit "name" tuple for naming the (true, false) resulting datasets. Note that internally, this creates 2 separate dataframes to represent each resulting group. As such, it is more of a shorthand function of doing such processing manually

```python
compare_internal_column = sv.compare_intra(iris,iris['Species'] == 'Iris-setosa',['Setosa','Other'])
compare_internal_column.show_html('internal_column.html')
```

```python
IFrame(src = 'internal_column.html',width=1000,height=500)
```
