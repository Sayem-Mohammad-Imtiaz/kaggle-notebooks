```python

plt.figure(figsize = (10,7))
chains = df_fiction['Author'].value_counts()[:11] #top 11
sns.barplot(x = chains, y = chains.index)
plt.title("Most popular Authors (Fiction)")
plt.xlabel("Count")
```

```python

plt.figure(figsize = (10,7))
chains = df_nonFiction['Author'].value_counts()[:9] #top 9
sns.barplot(x = chains, y = chains.index)
plt.title("Most popular Authors (Non Fiction)")
plt.xlabel("Count")
```
