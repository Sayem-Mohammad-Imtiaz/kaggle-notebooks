```python
import pandas as pd
df = pd.read_csv("../input/calcofi/bottle.csv")
df.head(8)
```

```python
df2 = df[['Salnty', 'T_degC']]
```

```python
df2.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df2.dropna(axis = 0,inplace = True)
```

```python
df2.shape
```

```python
df2.isnull().sum()
```

```python
df2 = df2[:][:500]
```

```python
X = df2['Salnty'].values.reshape(-1,1)
y = df2['T_degC'].values.reshape(-1,1)
```

```python
X=scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y,train_size = .75, random_state = 0)
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
pred = lr.predict(X_test)
pred
```

```python
predictions = scaler.inverse_transform(pred)
predictions
```

```python
lr.score(X_test, y_test)
```

```python
from xgboost import XGBRegressor
model = XGBRegressor(n_estimators = 500, learning_rate = 0.4, max_depth = 10)
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
