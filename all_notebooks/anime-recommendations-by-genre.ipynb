# Anime Recommendations

### Analyzing the Data

#### Looking at the ratings dataset, there are 3 columns mentioned user_id, anime_id and ratings.
First I want to remove the anime where the user has not given any rating even after watching the anime (-1).
This is under the assumption that the user did not find this anime appropriate or was not bothered to rate the anime.
However, this assumption will skip the fact that some users are not bothered to provide ratings. 

```python
import numpy as np
import pandas as pd
import itertools
import collections
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import jaccard_similarity_score 
from mpl_toolkits.mplot3d import Axes3D
```

```python
rating = pd.read_csv('../input/rating.csv')
print (rating.head())
```

```python
rating_clean = rating.drop(rating[rating.rating == -1].index)
print (rating_clean.head())

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs=rating_clean['user_id'], ys=rating_clean['anime_id']
           , s=20, c=None, depthshade=True)#, *args, **kwargs)
#ax.scatter(xs=rating_clean['rating'], ys=rating_clean['anime_id'])#, *args, **kwargs)
plt.show()
    
```
