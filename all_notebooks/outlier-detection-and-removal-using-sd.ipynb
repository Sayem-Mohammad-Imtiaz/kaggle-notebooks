```python
# Required Libraries
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Loading data into Data Frame
df = pd.read_csv('/kaggle/input/gender-and-height-dataset/gender-height.csv')
```

```python
df.head()
```

```python
# Shape of Data Set
df.shape
```

```python
# Plotting Histogram on Heights
plt.hist(df.Height, bins=20, rwidth = 0.8)
plt.xlabel('Heights(inches)')
plt.ylabel('Counts')
plt.show()
```

```python
# Upper limit of Heights
upperlimit = df.Height.mean() + 3*df.Height.std()
upperlimit
```

```python
# Lower limit of Heights
lowerlimit = df.Height.mean() - 3*df.Height.std()
lowerlimit
```

```python
# Outliers 
df[(df.Height > upperlimit)|(df.Height < lowerlimit)]
```

```python
# Removing Outliers 
df_new = df[(df.Height < upperlimit)&(df.Height > lowerlimit)]
```

```python
# shape of new data frame
df_new.shape
```

```python
# No of Outliers removed
df.shape[0] - df_new.shape[0]
```
