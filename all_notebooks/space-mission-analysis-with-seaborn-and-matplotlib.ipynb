# Launch year

First i wrote a function to fetch the year from datum feature then checking the top 10 most rocket launched years,as shown below 1917 tops the list.

```python
def get_year(x):
    return x[12:16]
data['Year']=data['Datum'].map(get_year)
data['Year']=data['Year'].astype('int64')
data['Year'].value_counts()[:10].plot(kind='bar',color='brown')
```

Now checking most successful rocket launched years,as shown below 2018 tops the list.

```python
Success=data[data['Status Mission']=='Success']
Success['Year'].value_counts()[:10].plot(kind='bar',color='brown')
```

#  Launch Month

Here also we are creating a new feature called month and checking the rockets launched per month,as shown below december tops the list.

```python
def get_month(x):
    return x[4:7]
data['Month']=data['Datum'].map(get_month)
sns.countplot(data['Month'])
```

Now we are checking most successful rocket launched months as shown below december tops the list.

```python
Successs=data[data['Status Mission']=='Success']
Successs['Month'].value_counts()[:10].plot(kind='bar',color='yellow')
```
