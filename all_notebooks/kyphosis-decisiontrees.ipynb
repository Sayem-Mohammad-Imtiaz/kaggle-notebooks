```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
kyphosis = pd.read_csv("../input/kyphosis-dataset/kyphosis.csv")
kyphosis
```

```python
sns.countplot(kyphosis['Kyphosis'], label = 'Count')
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
lc = LabelEncoder()
kyphosis['Kyphosis'] = lc.fit_transform(kyphosis['Kyphosis'])
```

```python
kyphosis
```

```python
kyphosis_true = kyphosis[kyphosis['Kyphosis'] == 1]
kyphosis_true
```

```python
kyphosis_false = kyphosis[kyphosis['Kyphosis'] == 0]
kyphosis_false
```

```python
print("Disease present after operation percentage = ", ( len(kyphosis_true)/len(kyphosis))*100)
```

```python
sns.heatmap(kyphosis.corr(), annot = True)
```

```python
sns.pairplot(kyphosis, hue = 'Kyphosis', vars = ['Age', 'Number', 'Start'])
```

```python
X = kyphosis.drop(['Kyphosis'], axis = 1)
X
```

```python
y = kyphosis['Kyphosis']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
dt.feature_importances_
```

```python
feature_importances = pd.DataFrame(dt.feature_importances_, index = X_train.columns, columns = ['importance'])
feature_importances
```

```python
feature_importances = pd.DataFrame(dt.feature_importances_, index = X_train.columns, columns = ['importance']).sort_values('importance', ascending = False)
feature_importances
```

```python
from sklearn.metrics import confusion_matrix, classification_report
y_pred = dt.predict(X_train)
cm = confusion_matrix(y_train, y_pred)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_train, y_pred))
```

```python
y_pred_test = dt.predict(X_test)
cm = confusion_matrix(y_test, y_pred_test)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_pred_test))
```

## improving the model

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators = 150)
rfc.fit(X_train, y_train)
```

```python
y_pred = rfc.predict(X_train)
cm = confusion_matrix(y_train, y_pred)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_train, y_pred))
```

```python
y_pred_test = rfc.predict(X_test)
cm = confusion_matrix(y_test, y_pred_test)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(y_test, y_pred_test))
```
