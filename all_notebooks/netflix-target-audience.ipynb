```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
netflix_filepath = "../input/netflix-shows/netflix_titles.csv"

netflix_data = pd.read_csv(netflix_filepath)
netflix_ratings =netflix_data['rating'].value_counts()
```

```python
#netflix_data.head()
#netflix_data
netflix_ratings
```

number of shows available on Netflix = 7669
number of shows suitable for children 13 and under = 2229

* TV-MA=15+
* TV-14=14+
* TV-PG=PG(<13)
* TV-Y =for all ages
* TV-Y7=for children above 7
* TV-G =for all ages
* NR   = not rated 
* TV-Y7-FV=for children 7 and above (include fantsy violence)
* UR   =Unrated
* NC-17=not for children below 17


What proportioned of the available shows are for younger viewers (PG and below)?

```python
plt.figure(figsize=(14,8))
sns.barplot(x=netflix_ratings.index, y=netflix_ratings)
plt.title("Ratings and Viewership")
```
