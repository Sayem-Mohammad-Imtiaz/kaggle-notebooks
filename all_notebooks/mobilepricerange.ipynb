```python
import pandas as pd
import numpy as np 
import seaborn as sns 
import matplotlib
```

```python
train = pd.read_csv('../input/mobile-price-classification/train.csv')
test = pd.read_csv('../input/mobile-price-classification/train.csv')
```

```python
print(train.shape)
print(test.shape)
```

```python
train.corr()
```

**We need to notice the feature which are very lightly corelated with the price Range**
***We will remove those feature as they aren't affecting much of the datsets***

**clock_speed** **m_dep**  **n_cores**

```python
del train['clock_speed'] 
del train['m_dep']
del train['n_cores']
del test['clock_speed'] 
del test['m_dep']
del test['n_cores']
```

```python
train.head()
```

```python
print(train.isnull().sum())
print(test.isnull().sum())
```

**FEATURE SCALING**

```python
from sklearn.preprocessing import StandardScaler
sts = StandardScaler()
feature_scale = ['battery_power','int_memory','mobile_wt','px_height','px_width','ram']
train[feature_scale] = sts.fit_transform(train[feature_scale])
test[feature_scale] = sts.fit_transform(test[feature_scale])
```

```python
train.head()
```

**IMPORTING MODELS**

```python
x = train.drop('price_range',axis =1 )
y = train['price_range']
ntest = test.drop('price_range',axis = 1)
```

```python
from sklearn import svm
```

```python
clf = svm.SVC()
```

```python
clf.fit(x,y)
```

```python
answer = clf.predict(ntest)
```

```python
from sklearn.metrics import r2_score
r2_score(test['price_range'],answer)
```

```python
result = pd.DataFrame({'PricePredict':answer})
result.to_csv('answer.csv',index = False)
```

**Hence our model prection is of 86 percent**
