### Plot to understand the overall diet preference in India

```python
plt.figure(figsize=(10,8))
plt.title("Count of veg/non-veg dishes")
sns.countplot(ind_food['diet'])
plt.show()
```

### **Inference**
1. Vegetarian dishes are more popular than non-veg dishes

### understanding the cooking time for the veg/non-veg dishes categorised based on the states

```python
plt.figure(figsize=(10,15))
plt.title("state vs cooking time for veg/non veg")
sns.barplot(x ='cook_time', y ='state', data = ind_food, hue='diet') 
plt.show()
```

### **Inference**
1. Highest cooking time by states is in telangana (no-veg) dish
2. Overall non veg dishes take more time to get cooked
3. The cooking time for vegetarian dishes are relatively less while in Bihar it is an exception.

```python
sns.catplot(x ='cook_time', y ='region', data = ind_food, hue='diet', col='flavor_profile', kind="bar",
                height=6, aspect=.4) 
```

### **Inference**
1. The sweet dishes falls under the veg diet with max cooking time in western region of India
2. The spicy flavor profile has a good mix of veg/non - veg diets with south India having the max cooking time
3. Bitter & Sour is relatively less popular and has veg diet only
4. few of the flavor pallets are unknown representated by -1 and also has a mix of veg/non-veg diets

```python
plt.figure(figsize=(15,10))
plt.title("Categorisation of courses based on diet preferences and preparation time")
sns.swarmplot(data=ind_food, y='course',x='prep_time', hue='diet')
plt.show()
#Categorical scatterplot with non-overlapping points
```

### **Inference**
1. Desert has the highest preference under veg
2. The main course has a good blend of veg and non veg diet and has the longest overall prep time
3. snacks are veg meals with maximum prep time of 1-100 minutes
