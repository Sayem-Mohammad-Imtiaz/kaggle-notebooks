# **Time Series Analysis on closing Stock and Volume**

```python
close = df[['Close']]
volume2 = df[['Volume']]
```

```python
def plot_timeseries(tdf,label1,label2,col,title):
    plt.figure(figsize=(25,15))
    sns.lineplot(data=tdf,label=label1,legend=False,linewidth=1)
    sns.lineplot(data=df[col],label=label2,legend=False,linewidth=1)
    plt.title(title)
    plt.legend()
    plt.show()
```

# **Simple Moving Average**

```python
mv_Close = close.rolling(window=30).mean()
plot_timeseries(mv_Close,'Moving Average','Actual','Close','Closing stock Moving Average')
```

```python
mv_Volume = volume2.rolling(window=30).mean()
plot_timeseries(mv_Volume,'Moving Average','Actual','Volume','Stock Volume Moving Average')
```

# **Weighted Moving Average**

```python
weights = np.arange(1,31)
weightedMV = df['Close'].rolling(30).apply(lambda close: np.dot(close, weights)/weights.sum(), raw=True)

plot_timeseries(weightedMV,'Weighted Moving Average','Actual','Close','Closing stock Weighted Moving Average')
```

```python
weights = np.arange(1,31)
volumeMV = df['Volume'].rolling(30).apply(lambda close: np.dot(close, weights)/weights.sum(), raw=True)

plot_timeseries(volumeMV,'Weighted Moving Average','Actual','Volume','Closing stock Weighted Moving Average')
```

# **Exponential Moving Average**

```python
exClose = df['Close'].ewm(span=30, adjust=False).mean()

plot_timeseries(exClose,'Exponential Moving Average','Actual','Close','Closing stock Exponential Moving Average')
```

```python
exVolume = df['Volume'].ewm(span=30, adjust=False).mean()

plot_timeseries(exVolume,'Exponential Moving Average','Actual','Volume','Volume stock Exponential Moving Average')
```
