```python
# filling missing values
def fill_missing_values(df):
    for i in df.columns:
        if i!='Transmission Classification' and df[i].isna().sum()>0:
            df[i].fillna(value = df[i].mean(),inplace = True)
fill_missing_values(df)

print(df.isna().sum().sum())
```
