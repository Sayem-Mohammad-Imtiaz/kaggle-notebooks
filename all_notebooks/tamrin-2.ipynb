```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../input/hotel-booking/hotel_booking.csv")
df.head()
```

```python
df = df.drop('company', axis = 1)
df.head(10)
```

```python
df.columns
```

```python
df['passangers']=df['adults']+df['children']+df['babies']

sf = df[['country','passangers']].groupby('country').sum()
sf.sort_values('passangers', ascending=False).head()
```

```python
kf = df[['arrival_date_month', 'passangers']].groupby('arrival_date_month').sum()
plt.figure(figsize=(14,6))
plt.xlabel('Months')
sns.histplot(data=df, x='arrival_date_month', hue='arrival_date_month')
```

**According to the graph we can understand that people mostly take a vacation in July and August. So hotels shold provide thier best serivces in these month.**

```python
df['meal'].value_counts()
sns.countplot(data=df, x='meal', hue='customer_type')
```
