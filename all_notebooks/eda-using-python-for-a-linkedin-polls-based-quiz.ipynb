```python
"""
## EDA Using SweetViz Library
"""

file_name='../input/linkedin-poll-data/LIQuizPollData%20-%20Sheet1%20(2).csv'
import pandas as  pd
df = pd.read_csv(file_name)
# importing sweetviz
import sweetviz as sv
#analyzing the dataset
quiz_report = sv.analyze(df)
#display the report
quiz_report.show_html('LIPollQuiz.html')
df1 = sv.compare(df[51:], df[:51])
df1.show_html('LIQuizPollCompare.html')
```

```python
"""## EDA Using AutoViz Library"""
from autoviz.AutoViz_Class import AutoViz_Class
AV = AutoViz_Class()
file_name='https://bit.ly/3BohVHu'
df = AV.AutoViz(file_name)
"""## EDA Using lux Library"""
import lux
import pandas as pd
df = pd.read_csv(file_name)
df
df.intent = ["Total_views","Total_Responses"]
df
```
