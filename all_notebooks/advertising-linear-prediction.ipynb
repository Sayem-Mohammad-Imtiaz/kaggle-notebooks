```python
# Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
# Import data
advert = pd.read_csv("../input/advertising-dataset/advertising.csv")
```

```python
advert.head()
```

```python
# Check null values
advert.isnull().sum()
```

```python
# More information, data types
advert.info()
```

```python
advert.describe()
```

```python
# Graphs
sns.pairplot(advert)
```

```python
#Data distribution of each variable
plt.figure(figsize=(9,6))
sns.distplot(advert['TV'])
```

```python
plt.figure(figsize=(9,6))
sns.distplot(advert['Radio'])
```

```python
plt.figure(figsize=(9,6))
sns.distplot(advert['Newspaper'])
```

```python
# Dividing data in train-test
X = advert[['TV','Radio','Newspaper']].values
y = advert['Sales'].values
```

```python
print(X.shape)
print(y.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
# Define model
linereg = LinearRegression()
```

```python
#Fitting the model
linereg.fit(X_train,y_train)
```

```python
#coefficients of model
linereg.coef_
```

```python
# Intercept
linereg.intercept_
```

```python
# Predicting 
y_predict = linereg.predict(X_test)
```

```python
# Prediction
y_predict
```

```python
# Model Evaluation
print("Mean Abs Error :",metrics.mean_absolute_error(y_test,y_predict))
print("Mean Sq Error  :",metrics.mean_squared_error(y_test,y_predict))
print("Root Mean Sq Error :",np.sqrt(metrics.mean_squared_error(y_test,y_predict)))
print("r2 value   :",metrics.r2_score(y_test,y_predict))
```
