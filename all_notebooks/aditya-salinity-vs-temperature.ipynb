# **Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# **Importing Data**

```python
df=pd.read_csv('../input/calcofi/bottle.csv')
df.head(8)
```

# **Creating Dependent and Independent Vectors**

```python
X=df.iloc[:,5].values
Y=df.iloc[:,6].values
```

```python
print(X)
```

```python
print(Y)
```

# **Reshaping arrays**

```python
X=np.reshape(X,(864863,1))
Y=np.reshape(Y,(len(Y),-1))
```

```python
print(X)
```

```python
print(Y)
```

> # **Checking For Missing Values**

```python
print(np.any(np.isnan(X)))  #If true Missing Value available
print(np.any(np.isnan(Y)))
```

# **Removing missing values with mean**

```python
from sklearn.impute import SimpleImputer as sp
imputer=sp(missing_values=np.nan,strategy='mean')
imputer.fit(X)
imputer.fit(Y)
X=imputer.transform(X)
Y=imputer.transform(Y)
```

```python
print(X)
```

```python
print(Y)
```

```python
Y=np.reshape(Y,(-1))
```

```python
print(Y)
```

# **Splitting into training and test set**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
print(Y_train)
```

```python
print(Y_test)
```

# **Feature Scaling**

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
print(X_train)
print(X_test)
```

# **Training The Model**

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)

```

# **Predicting Result**

```python
Y_pre=lr.predict(X_train)
print(Y_pre)
```

# **Showing Results with plots**

```python
plt.scatter(X_train,Y_train,color='green')
plt.plot(X_train,Y_pre,color='red')
plt.title('Temperature Vs Salinity')
plt.xlabel('Temperature')
plt.ylabel('Salinity')
plt.show()
```

```python
plt.scatter(X_test,Y_test,color='green')
plt.plot(X_train,Y_pre,color='red')
plt.title('Temperature Vs Salinity')
plt.xlabel('Temperature')
plt.ylabel('Salinity')
plt.show()
```
