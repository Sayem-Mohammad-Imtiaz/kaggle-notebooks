# **India**

```python
Agriculture=pd.read_csv("../input/world-population-working-in-different-sectors/Agriculture.csv")
```

```python
Agriculture.head()
```

```python
India=Agriculture[Agriculture['Country Name']== 'India']
India=India.drop(['Country Code','2020'], axis = 1)
```

```python
india_Agriculture=India.set_index('Country Name').T.astype(float)
india_Agriculture.plot(figsize=(20,10),marker="o",title="% of total population in agriculture from 1991 to 2019")
plt.xlabel("YEARS")
plt.ylabel("% of total population in agriculture")
plt.legend()
plt.show()
```

```python
Industry=pd.read_csv("../input/world-population-working-in-different-sectors/Industry.csv")
```

```python
India=Industry[Industry['Country Name']== 'India']
India=India.drop(['Country Code','2020'], axis = 1)
```

```python
india_Industry=India.set_index('Country Name').T.astype(float)
india_Industry.plot(figsize=(20,10),marker="o",title="% of total population in Industry from 1991 to 2019")
plt.xlabel("YEARS")
plt.ylabel("% of total population in Industry")
plt.legend()
plt.show()
```

```python
Service=pd.read_csv("../input/world-population-working-in-different-sectors/Service.csv")
```

```python
India=Service[Service['Country Name']== 'India']
India=India.drop(['Country Code','2020'], axis = 1)
```

```python
india_Service=India.set_index('Country Name').T.astype(float)
india_Service.plot(figsize=(20,10),marker="o",title="% of total population in Service from 1991 to 2019")
plt.xlabel("YEARS")
plt.ylabel("% of total population in Service")
plt.legend()
plt.show()
```

# **LinearRegression Model**

```python
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
```

```python
india_Agriculture=india_Agriculture.values
```

```python
india_Industry=india_Industry.values
```

```python
lin.fit(india_Industry,india_Agriculture)
```

```python
y_pred=lin.predict(india_Industry)
```

```python
plt.scatter(india_Industry,india_Agriculture)
plt.plot(india_Industry,y_pred,color='red')
plt.xlabel('Industry Population')
plt.ylabel('Agriculture Population')
plt.title('Agriculture Population v/s Industry Population')
plt.show()
```

```python
india_Service=india_Service.values
```

```python
lin.fit(india_Service,india_Agriculture)
```

```python
plt.scatter(india_Service,india_Agriculture)
plt.plot(india_Service,y_pred,color='red')
plt.xlabel('Service Population')
plt.ylabel('Agriculture Population')
plt.title('Agriculture Population v/s Service Population')
plt.show()
```

```python
df_A= Agriculture[(Agriculture["Country Name"]=="Bhutan") |(Agriculture["Country Name"]=="China")|(Agriculture["Country Name"]=='Nepal') |(Agriculture["Country Name"]=='Afghanistan')|(Agriculture["Country Name"]=='India')|(Agriculture["Country Name"]=='Pakistan')|(Agriculture["Country Name"]=='Bangladesh')|(Agriculture["Country Name"]=='Myanmar')]
```

```python
import plotly.express as px
```

```python
Agriculture_map = df_A.set_index('Country Name')
temp = pd.DataFrame(Agriculture_map['2019']).reset_index()


temp['iso_alpha'] = temp['Country Name'].apply(lambda x:pc.country_name_to_country_alpha3(x,))
fig = px.choropleth(temp, locations='iso_alpha',
                    color='2019',
                    hover_name='Country Name',
                    color_continuous_scale=px.colors.diverging.RdYlGn,
                   )
fig.update_layout(
    title_text='World map - 2019',
    showlegend=False,
    paper_bgcolor='rgb(248, 248, 255)',
    geo_bgcolor='rgb(248, 248, 255)',
    geo_showframe=False,
    title_font_size=22,
)
fig.show()
```
