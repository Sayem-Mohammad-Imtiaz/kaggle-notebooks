# Deccan Plateau

```python
options = ['Madhya Maharashtra', 'Matathwada', 'Vidarbha', 'North Interior Karnataka', 
           'South Interior Karnataka']
df_dp = df[df["SUBDIVISION"].isin(options)]
df_dp.head()
```

```python
df_dp["WINTER"] = df_dp["DEC"] + df_dp["JAN"]
df_dp["SPRING"] = df_dp["FEB"] + df_dp["MAR"]
df_dp["SUMMER"] = df_dp["APR"] + df_dp["MAY"] + df_dp["JUN"]
df_dp["MONSOON"] = df_dp["JUL"] + df_dp["AUG"] + df_dp["SEP"]
df_dp["AUTUMN"] = df_dp["OCT"] + df_dp["NOV"]
```

```python
df_dp_melt = pd.melt(df_dp, id_vars = ["SUBDIVISION", "YEAR"], value_vars = ["WINTER", "SPRING",
                                                                    "SUMMER", "MONSOON", "AUTUMN"])
df_dp_melt1 = df_dp_melt.groupby(["SUBDIVISION", "variable"])["value"].sum().reset_index()
```

```python
fig = px.bar(df_dp_melt1, labels={"variable": "Seasons"},
              x = "SUBDIVISION", y = "value", color = "variable", facet_row = "variable",
            color_discrete_map= {
                "WINTER":"#000080",
                "SPRING":"#FF69B4",
                "AUTUMN":"#CC5500",
                "SUMMER":"#FF9933",
                "MONSOON":"#C2B280",
            })

fig.for_each_annotation(lambda a: a.update(text=a.text.split("=")[-1]))

# hide subplot y-axis titles and x-axis titles
for axis in fig.layout:
    if type(fig.layout[axis]) == go.layout.YAxis:
        fig.layout[axis].title.text = ''
    
        
# keep all other annotations and add single y-axis and x-axis title:
fig.update_layout(
    plot_bgcolor = "#EEEEEE",
    paper_bgcolor = "#f6f5f5",
    autosize = True,
    xaxis = dict(
        title_text = "<b> States </b>",
        titlefont = dict(size = 12)
    ),
    
    title_text = "<b> Seasonal Distribution of Rainfall in Deccan Plateau</b>",
    title_font_size = 16,
    title_font_color = "black",
    title_pad_t = 5,
    title_pad_l = 20,
    # keep the original annotations and add a list of new annotations:
    annotations = list(fig.layout.annotations) + 
    [go.layout.Annotation(
            x=-0.07,
            y=0.5,
            font=dict(
                size=12
            ),
            text="<b> Annual Rainfall </b>",
            textangle=-90,
            xref="paper",
            yref="paper"
        )
    ] 
)
    

fig.update_yaxes(showticklabels = True, showline = True, linewidth = 2, linecolor = "black",  matches=None)
fig.update_xaxes(showline = True, linewidth = 2, linecolor = "black")

py.offline.iplot(fig)
```

From above we can see that spring has the least rainfall in all states while monsoon has maximum rainfall. Among all the states above Vidarbha has the maximum rainfall in Deccan Plateau.
