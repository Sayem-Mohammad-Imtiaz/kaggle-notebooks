```python
import numpy as np
import pandas as pd
import category_encoders as ce

from sklearn.metrics import classification_report, log_loss, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import ShuffleSplit

from xgboost import XGBClassifier
from tqdm import tqdm
```

```python
data = pd.read_csv("../input/video-game-tweets/videogame_tweets.csv")
data[0:2]
```

```python
data1=data[['clean_text','likes']]
data1[0:2]
```

```python
words=[]
for item in data1['clean_text']:
    for word0 in item[1:-1].split(', '):
        word=word0[1:-1]
        words+=[word]
print(len(words))
print(words[0:10])
```

```python
kw=pd.Series(words).value_counts()[0:100].index.values
kw
```

```python
kw0=np.array(['clean_text','likes'])
kw0
```

```python
kw2=np.concatenate([kw0,kw])
kw2
```

```python
data2=pd.DataFrame(data1,columns=kw2) 
data2[0:2]
```

```python
words3=[]
for item in data1['clean_text']:
    words2=[]
    for word0 in item[1:-1].split(', '):
        word=word0[1:-1]
        if word in kw:
            words2+=[word]
    words3+=[words2]
print(len(words3))
print(words3[0:10])
```

```python
data2[0:2]
```

```python
for i in tqdm(range(1135)):
    for j in range(2,102):
        if kw2[j] in data2['clean_text'][i]:
            data2.iloc[i,j]=1
        else:
            data2.iloc[i,j]=0
data2[0:2]
```

```python
n=len(data2)
print(n)
```

```python
train = data2[0:(n//10)*8]
test = data2[(n//10)*8:]
```

```python
y_train0 = train['likes']
y_test0 = test['likes']
X_train0 = train.drop(['clean_text','likes'],axis=1)
X_test0 = test.drop(['clean_text','likes'],axis=1)
```

```python
X = np.array(X_train0)
y = np.array(y_train0)
```

```python
clf = XGBClassifier(max_depth=3, n_estimators=1000, learning_rate=0.01)
```

```python
ss = ShuffleSplit(n_splits=5,train_size=0.8,test_size =0.2,random_state=0) 

for train_index, test_index in ss.split(X): 

    X_train, X_test = X[train_index], X[test_index]
    Y_train, Y_test = y[train_index], y[test_index]

    clf.fit(X_train, Y_train) 
    print(clf.score(X_test, Y_test))
```

```python
pred0 = clf.predict(np.array(X_test0))
```

```python
PRED=pred0
ANS=np.array(y_test0)
accuracy=accuracy_score(ANS,PRED)
print(accuracy)
```
