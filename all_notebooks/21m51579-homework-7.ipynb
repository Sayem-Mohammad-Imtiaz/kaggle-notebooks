# 1. Create a global map for COVID Vaccinations (Transition of daily vactinations per million all over the world).

In this global animation, I will show the transition of daily vacctinations per million all over the world. **With this map, it is able to compare the intensity of the vacctination policy of every country on that day period. Since I was curious to see where are the "hot spots" of the vacctination in the world on that time, I decided to visualize this column.** Therefore, from this map, we can easily discuss and evaluate the differences of the vacctination policies by each country.
(e.g., Since the redness of Japan is relatively weak all the time, Japanese vacctination policy seems delaying from global standard. e.g.2, Since the US has actively conducted the vactination policy, now, they passed the peak of the daily vacctinations.)

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly

df = pd.read_csv('../input/covid-world-vaccination-progress/country_vaccinations.csv', header=0)
df.index = pd.to_datetime(df['date'])
df = df.sort_index()
print(df.columns)
fig = px.choropleth(df, locations='country', locationmode='country names', color='daily_vaccinations_per_million', hover_name='country', animation_frame='date', color_continuous_scale='OrRd', range_color=(0.0, 10000.0))
fig.update_layout(title_text="Transition of daily vactinations per million all over the world", title_x=0.4)
fig.show()

plotly.offline.plot(fig,filename='vaccinations.html')
```

# 2. Create a global map of earthquake occurences worldwide.

In this hourly animation, I decided to visualize the record of October 19th, 2020, when a M7.6 earthquake struck off the Alaska Peninsula. We can recognize one red dot (an earthquake greater than M7) around 20:00 and many small earthquakes were happeding before and after the main earthquake around Alaska Peninsula. From the anaylsis and these news, I found that some severe earthquakes are happening in rural areas, and they are not broadcasted because they do not bring huge damage to our lives and properties. However, this dataset is grasping such a "hidden severe earthquake" and certainly enables us to visualize the geological impact (such as small quakes before and after the main quake).

```python
df = pd.read_csv('../input/20201019/query.csv', header = 0)
df.index = pd.to_datetime(df['time'])
df['time']= df.index.strftime('%Y-%m-%d %H:00:00')
df = df.sort_index()
fig = px.scatter_geo(df,lat='latitude', lon='longitude', color='mag', animation_frame='time',color_continuous_scale='rainbow', range_color=(1.0,7.0))
fig.show()
```
