```python
df.TIME = df.TIME.astype("str")
```

```python
#Create a new variable or column for month
df['Month'] = ""
Months = ['Jan-20','Feb-20','Mar-20','Apr-20','May-20','Jun-20','Jul-20','Aug-20','Sept-20','Oct-20']

month = []

for row in df.TIME:
    if row == '2020-01':
        month.append(Months[0])
    elif row == '2020-02':
        month.append(Months[1])
    elif row == '2020-03':
        month.append(Months[2])
    elif row == '2020-04':
        month.append(Months[3])
    elif row == '2020-05':
        month.append(Months[4])
    elif row == '2020-06':
        month.append(Months[5])
    elif row == '2020-07':
        month.append(Months[6])
    elif row == '2020-08':
        month.append(Months[7])
    elif row == '2020-09':
        month.append(Months[8])
    elif row == '2020-10':
        month.append(Months[9])
    else:
        month.append('Not Defined')

df.Month = month

df.tail(5)
```

```python
unemployment_data = df[['LOCATION','SUBJECT','TIME','Value','Month']]
unemployment_data = unemployment_data.rename(columns={'SUBJECT':'Gender','Value':'Unemployment Rate'})
```

```python
#unique subjects
unemployment_data.Gender.unique()
```

```python
#datasets by gender
List_1 = ['TOT']
List_2 = ['MEN']
List_3 = ['WOMEN']

unemployment_total = unemployment_data[unemployment_data.Gender.isin(List_1)]
unemployment_men = unemployment_data[unemployment_data.Gender.isin(List_2)]
unemployment_women = unemployment_data[unemployment_data.Gender.isin(List_3)]
```

```python
unemployment_women.Gender.unique()
```

```python
#import the plotly packages
import plotly.offline as py
from plotly import tools
import plotly.graph_objs as go
import plotly.express as px
from plotly.subplots import make_subplots
```

```python
#unemployment for each country in during Covid19
#North America
#total unemployment, unemployment among men, unemployment among women

fig1 = px.choropleth(unemployment_total,locations=unemployment_total['LOCATION'],color=unemployment_total['Unemployment Rate'],color_continuous_scale='Reds',locationmode='ISO-3',scope='north america',animation_frame=unemployment_total['Month'],title="Total unemployment during Covid-19",labels={'color':'Unemployment Rate'})
fig2 = px.choropleth(unemployment_men,locations=unemployment_men['LOCATION'],color=unemployment_men['Unemployment Rate'],color_continuous_scale='Reds',locationmode='ISO-3',scope='north america',animation_frame=unemployment_men['Month'],title="Men: Unemployment during Covid-19",labels={'color':'Unemployment Rate'})
fig3 = px.choropleth(unemployment_women,locations=unemployment_women['LOCATION'],color=unemployment_women['Unemployment Rate'],color_continuous_scale='Reds',locationmode='ISO-3',scope='north america',animation_frame=unemployment_women['Month'],title="Women: Unemployment during Covid-19",labels={'color':'Unemployment Rate'})

fig1.show()
fig2.show()
fig3.show()
```

```python
#unemployment for each country in during Covid19
#Europe
#total unemployment, unemployment among men, unemployment among women

fig4 = px.choropleth(unemployment_total,locations=unemployment_total['LOCATION'],color=unemployment_total['Unemployment Rate'],color_continuous_scale='Reds',locationmode='ISO-3',scope='europe',animation_frame=unemployment_total['Month'],title="Total unemployment during Covid-19",labels={'color':'Unemployment Rate'})

fig4.show()
```
