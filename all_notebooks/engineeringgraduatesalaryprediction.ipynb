```python
df['Age']=((pd.to_datetime('today') - pd.to_datetime(list(df['DOB']))).days / 365).astype(int)
df
```

```python
df=df.drop([
 '10board',
 '12board',
 'ID',
 '12graduation',
 'CollegeID',
 'CollegeTier',
 'CollegeCityID',
 'CollegeCityTier',
 'GraduationYear','CollegeState','DOB','conscientiousness','agreeableness','extraversion','nueroticism','10percentage','12percentage',
    'ComputerProgramming','ElectronicsAndSemicon','ComputerScience','MechanicalEngg','ElectricalEngg','TelecomEngg','CivilEngg'                
],axis=1)
```

```python
df
```

```python
#Gender 0-Female 1-Male
```

```python
df['Gender'] = df['Gender'].map({'f': 0,'m': 1})
```

```python
data = pd.get_dummies(df, drop_first =True)
data.head()
```

```python
data.info()
```

```python
upper_tri=data.corr()
upper_tri
```

```python
#to_drop = [column for column in upper_tri.columns if any((upper_tri[column] < 0.5) & (upper_tri[column]< 0.9 ))]
to_drop = [column for column in upper_tri.columns if any((upper_tri[column] < 0.1) )]
to_drop
```

```python
data=data.drop('Domain',axis=1)
```

```python
X = data.drop('Salary' , axis = 1)
y=data['Salary']
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split( X,y , test_size = 0.3, random_state = 0) 
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(X_train, y_train)
reg.score(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

MSE  = mean_squared_error(y_test, y_pred)
print("MSE :" , MSE)

RMSE = np.sqrt(MSE)
print("RMSE :" ,RMSE)
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R2 :" ,r2)
```

```python
print('Train Score: ', reg.score(X_train, y_train))  
print('Test Score: ', reg.score(X_test, y_test))  
```

```python
from sklearn import linear_model
lasso  = linear_model.Lasso(alpha=1 , max_iter= 3000)

lasso.fit(X_train, y_train)
```

```python
lasso.score(X_train, y_train)
```

```python
y_pred_l = lasso.predict(X_test)
```

```python
MSE  = mean_squared_error(y_test, y_pred_l)
print("MSE :" , MSE)

RMSE = np.sqrt(MSE)
print("RMSE :" ,RMSE)

r2 = r2_score(y_test, y_pred_l)
print("R2 :" ,r2)
```

```python
from sklearn.linear_model import Ridge

ridge  = Ridge(alpha=0.1)
ridge.fit(X_train,y_train)
```

```python
y_pred_r = ridge.predict(X_test)
MSE  = mean_squared_error(y_test, y_pred_r)
print("MSE :" , MSE)

RMSE = np.sqrt(MSE)
print("RMSE :" ,RMSE)

r2 = r2_score(y_test, y_pred_r)
print("R2 :" ,r2)
```
