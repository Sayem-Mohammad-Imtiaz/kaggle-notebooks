# A little detail about what each column means

- **Date** : Event Date
- **Time** : Event Time
- **Country** : Country of the Event
- **Votality** : The degree of volatility (possible fluctuations in currency, indices, etc.) caused by this event.
- **Description** : Description of the event
- **Evaluation** : Evaluation of the event according to the actual data, which came out better than the forecast, worse or correspond to it!
- **Percentage** : Data Format. (%, K x103, M x106, T x109)
- **Actual** : Actual Event Data
- **Forecast** : Event Forecast Data
- **Previous Data** : Previous data on this event (with comments if there were any interim changes)

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Checking out the events epicentre maxima around the world

```python
plt.figure(figsize=(15,8))
sns.barplot(y = df['Country'].value_counts().index, x = df['Country'].value_counts(), palette = 'summer')
plt.title("Epicenters of Forex events around the world")
plt.xlabel('Number of Events')
```

## Note : 

1. As it is evident, even in Pre and Pro **Brexit Era** combined, the Euro Zone doesn't even seem to be a close competitor of the Bullish Market of the United States.      

2. Looking closely at the **Asian** market, one can easily judge that it's not really doing anything terrific , if talking in terms of magnitude of such decisions taking place. 

3. Also, another thing worth noticing is the seperate mention of **strong EU Members** like Germany, France and Spain (Okay, not really strong, but still noteworthy)
