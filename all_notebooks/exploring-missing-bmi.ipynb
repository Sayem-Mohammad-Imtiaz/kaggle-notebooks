# MAKING MODEL WITH MEDIAN VALUE INSTEAD OF NULL VALUES IN BMI

```python
df_m = df.copy()
```

```python
df_m['bmi'] = df_m['bmi'].fillna(28.1)
```

```python
df_m.isnull().any()
```

```python
df_m.head()
```

Converting Strings into Integers
1. GENDER

2.Ever Married

3.Work Type

4.Residence Type

5.Smoking Status

There are furthur two ways to either use dummy varibales or do label encoding , first we would go for label encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df_m['gender'] = le.fit_transform(df_m['gender'])
df_m['ever_married'] = le.fit_transform(df_m['ever_married'])
df_m['work_type'] = le.fit_transform(df_m['work_type'])
df_m['Residence_type'] = le.fit_transform(df_m['Residence_type'])
df_m['smoking_status'] = le.fit_transform(df_m['smoking_status'])
```

```python
df_m.head(2)
```

```python
df.columns
```

```python
from sklearn.model_selection import train_test_split
x = df_m[['gender', 'age', 'hypertension', 'heart_disease', 'ever_married',
       'work_type', 'Residence_type', 'avg_glucose_level', 'bmi',
       'smoking_status']]
y = df_m['stroke']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,stratify=y,random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
log.fit(x_train,y_train)
```

```python
log.score(x_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
sns.heatmap(confusion_matrix(y_test,log.predict(x_test)),annot=True)
    
```
