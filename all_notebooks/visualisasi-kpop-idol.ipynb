# **Buat Line Chart, Histogram dan Scatterplot**

```python
boy = pd.read_csv("../input/kpopdb/kpop_idols_boy_groups.csv")
```

```python
boy
```

```python
boy.info()
```

```python
#Linechart
import matplotlib.pyplot as plt
  
plt.plot(boy['Members'], boy['Orig. Memb.'], color='green', marker='o')
plt.title('Members Vs Orig. Memb.', fontsize=14)
plt.xlabel('Members', fontsize=14)
plt.ylabel('Orig. Memb.', fontsize=14)
plt.grid(True)
plt.show()
```

```python
#Histogram
plt.hist(boy['Members'], bins=9, align='right', color='blue', edgecolor='white')
```

```python
#Scatterplot
import seaborn as sns
sns.scatterplot(x=boy['Members'], y =boy['Orig. Memb.'], hue = boy['Active'])
```

# **Pair plot, Styling dan Boxplot**

```python
sns.pairplot(boy)
```

```python
#styling
sns.set_style('white')

# You can use figure sizing from Matplotlib
plt.figure(figsize=(8,4))

# Change size of lables, lines and other elements to best fit
# how you will present your data (paper, talk, poster)
sns.set_context('paper', font_scale=1.4)

sns.jointplot(x='Members', y='Orig. Memb.', data=boy, kind='reg')

# Get rid of spines
# You can turn of specific spines with right=True, left=True
# bottom=True, top=True
sns.despine(left=False, bottom=False)
```

```python
#boxplot
sns.boxplot(x="Members", y="Orig. Memb.", data=boy)
```
