# Train And Test <a id="2"></a>

Using all other columns to predict the sex.

```python
X = df[['chol', 'cp', 'trestbps', 'age', 'fbs', 'restecg', 'thalach',
       'exang', 'oldpeak', 'slope', 'ca', 'thal', 'target']] .values  #.astype(float)
X[0:5]
```

As mentioned before I want to predict the sex.

```python
y = df['sex'].values
y[0:5]
```

Data Standardization give data zero mean and unit variance, it is good practice, especially for algorithms such as KNN which is based on distance of cases:

```python
X = preprocessing.StandardScaler().fit(X).transform(X.astype(float))
X[0:5]
```

Train and Test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
```

Just to get the best K value the code part below iterates 100 time to find the best K value

```python
Ks = 100
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))
print(mean_acc,"mean")
print(std_acc,"std")
```

```python
ConfustionMx = [];
for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

mean_acc
```

```python
plt.plot(range(1,Ks),mean_acc,'g')
plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
plt.legend(('Accuracy ', '+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Nabors (K)')
plt.tight_layout()
plt.show()
```

```python
print( "The best accuracy was with", mean_acc.max(), "with k=", mean_acc.argmax()+1)
```
