```python
import pandas as pd 
```

```python
data_df = pd.read_csv("/kaggle/input/emigrants-from-romania-19912020/emigrants_romania.csv")
```

```python
data_df.head()
```

```python
print("Emigration by sex")
data_df.groupby("sex")["total"].sum().reset_index().head()
```

```python
print("Emigration by region - top 5")
data_df.groupby("geo")["total"].sum().reset_index().sort_values(by="total", ascending=False).head()
```

```python
all_geo = list(data_df.geo.unique())
exclusion_list = [' TOTAL', ' MACROREGIUNEA UNU', ' Regiunea NORD-VEST',' Regiunea CENTRU', ' MACROREGIUNEA DOI', ' Regiunea NORD-EST', 
                  ' Regiunea SUD-EST', ' MACROREGIUNEA TREI',' Regiunea SUD-MUNTENIA', ' Regiunea BUCURESTI - ILFOV', ' MACROREGIUNEA PATRU',
                  ' Regiunea SUD-VEST OLTENIA', ' Regiunea VEST']
counties_only = list(set(all_geo) - set(exclusion_list))
print("Emigration by region - counties only - top 5")
data_df.loc[data_df.geo.isin(counties_only)].groupby("geo")["total"].sum().reset_index().sort_values(by="total", ascending=False).head()
```

```python
print("Emigration by year - top 5")
data_df.groupby("year")["total"].sum().reset_index().sort_values(by="total", ascending=False).head()
```
