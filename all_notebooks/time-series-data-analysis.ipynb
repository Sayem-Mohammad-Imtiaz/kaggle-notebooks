```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
%matplotlib inline
```

```python
df = pd.read_csv("../input/spotify-dataset-19212020-160k-tracks/data_o.csv")
df.shape
```

```python
df.drop(['explicit','id','key','mode','release_date'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.columns
```

```python
corel=df[['valence','acousticness','danceability','energy','instrumentalness', 'liveness', 'loudness','speechiness','tempo']].corr()
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(data=corel,annot=True)
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(data=corel,annot=True,cmap='tab20',linewidths=2,linecolor='white')
```

```python
Avarage_year = df[['valence','acousticness','danceability','energy','instrumentalness', 'liveness', 'loudness','speechiness','tempo','year']].groupby('year').mean().sort_values('year').reset_index()
```

```python
Avarage_year.head()
```

```python
Avarage_year.plot(figsize=(20,10),linewidth=1,fontsize=5)
plt.xlabel('year',fontsize=5);
```

```python
plt.figure(figsize=(20,10))
plt.title('Trend of Song over years',fontsize=15)
lines = ['acousticness','danceability','energy','instrumentalness','liveness','valence','speechiness']
for i in lines:
    sns.lineplot(x='year', y=i, data=Avarage_year)
plt.ylabel('value')
plt.legend(lines)
```

```python
plt.figure(figsize=(20,10))
plt.title('Energy factor of songs',fontsize=15)
sns.lineplot(x='year', y='energy', data=Avarage_year,linewidth=5)
```

```python
plt.figure(figsize=(20,10))
plt.title('Danceability factor of songs',fontsize=15)
sns.lineplot(x='year', y='danceability', data=Avarage_year,linewidth=5)
```
