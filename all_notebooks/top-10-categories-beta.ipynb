**Let us focus on the top 10 categories based on total amount pledged in USD (US Dollars)!**

```python
top10 = df.groupby('category').sum()[['converted currency(USD)','num.backers']].reset_index().sort_values('converted currency(USD)', ascending=False).head(10)

top10['num.backers'] = pd.to_numeric(top10['num.backers'].apply(lambda x: '%.0f' % x))
top10['converted currency(USD)'] = pd.to_numeric(top10['converted currency(USD)'].apply(lambda x: '%.0f' % x))
top10['avg. pledge per backer'] = top10['converted currency(USD)']/top10['num.backers']


top10
```

```python
top10_list = list(top10['category'])
top10_list
```

```python
plt.figure()

sns.barplot(x='category',y='converted currency(USD)',data=top10)
plt.tight_layout()
plt.ylabel('# Backers')
```
