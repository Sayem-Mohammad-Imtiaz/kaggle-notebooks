Based on the data visualized below, Starbucks has a very small presence in the southern hemisphere. The most densely populated markets in North America are New York City and Los Angeles.

```python
import pandas as pd
df = pd.read_csv('../input/directory.csv')
df.info() 
```

```python
df.head()
```

```python
#show the entire world
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks")
sns.set_context("talk")

x = df['Longitude']
y = df['Latitude']

sns.jointplot(x, y, kind="hex", size=10, color="#4CB391", stat_func=None, 
              xlim=(-180,180), ylim=(-180,180))
```

```python
countries = (df[["Country", "Store Number"]].groupby(["Country"], as_index=False).count())
countries.sort_values(by='Store Number', axis=0, ascending=False, inplace=True)
countries.head(25)
```

```python
countries = (df[["City", "Store Number"]].groupby(["City"], as_index=False).count())
countries.sort_values(by='Store Number', axis=0, ascending=False, inplace=True)
countries.head(25)
```

```python
#show North America
sns.jointplot(x, y, kind="hex", size=10, color="#4CB391", 
              xlim=(-130,-60), ylim=(10,60), stat_func=None)
```

The above jointplot suggests that Starbucks stores are equally distributed by longitude; that is they cover North America farily well from Pacific to Atlantic. The latitude distribution is somewhat normal and suggesting a peak across 35 degrees north.  Although the Gulf of Mexico may make this distribution a bit misleading, what's clear is that there are a lot of Mexican locations but only in very specific regions of the country (Mexico City is a top city as noted in the table above).

```python
#show Europe
sns.jointplot(x, y, kind="hex", size=10, color="#4CB391", 
              xlim=(-25,50), ylim=(25,65), stat_func=None)
```

```python
#show Asia
sns.jointplot(x, y, kind="hex", size=10, color="#4CB391", 
              xlim=(60,165), ylim=(-10,50), stat_func=None)
```
