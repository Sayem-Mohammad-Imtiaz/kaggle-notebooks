```python
teynampet = df[(df['block_name']=='Teynampet')].reset_index(drop=True)
teynampet.head(2)
```

#Code that shows the number of first doses available in Teynampet.

"And the color factor is used for a categorical feature from the dataset to see differences. Now I've used the feature 'allow_all_age' which is categorical with classes true and false. So it'll display whether all ages are allowed in that particular vaccination centre at Teynampet."

"Size - numerical attribute to quantify"

"Color - categorical attribute to classify the data points on the map according to the class of that categorical attribute."

Explanations by Varshini PJ.

```python
#Code by Varshini PJ  user: varshinipj

fig = px.scatter_mapbox(teynampet,
# Here, plotly gets, (x,y) coordinates
lat="lat",
lon="long",
text='block_name',

                #Here, plotly detects color of series
                size="available_capacity_dose1",
                color = "allow_all_age",
                labels="block_name",

                zoom=14.5,
                center={"lat":13.044403, "lon":80.251648},
                height=600,
                width=800)
fig.update_layout(mapbox_style='stamen-terrain')
fig.update_layout(margin={"r": 0, "t": 0, "l": 0, "b": 0})
fig.update_layout(title_text="Available first doses in Teynampet")
fig.show()
```
