```python
import plotly
import plotly.express as px
import numpy as np
import pandas as pd
from pandas.io.json import json_normalize
import math

rest_data = pd.read_csv('/kaggle/input/riyadh-restaurants-20k/riyadh_resturants_clean.csv', index_col=None)


p1 = {'lat': 24.719462, 'lon': 46.719108} # all boroghs
rest_data.loc[pd.isnull(rest_data['price']), 'price'] = 'Unknown'
fig = px.scatter_mapbox(rest_data,
                        lat=rest_data['lat'], 
                        lon=rest_data['lng'],  
                        #text=rest_data['location.fullName'],
                        center=p1, 
                        color=rest_data['price'],
                        zoom=9,
                        mapbox_style="open-street-map",
                        title='Riyadh Restaurants (~20K)'
                       )
fig.show()
```
