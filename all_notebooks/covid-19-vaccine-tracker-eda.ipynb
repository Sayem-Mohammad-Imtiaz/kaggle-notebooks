```python
cols_to_drop=['Funder','Published Results','Clinical Trials for Other Diseases (T only) / Related Use or Platform (V only)','FDA-Approved Indications','Sources']
df=df.drop(cols_to_drop,axis=1)
df.columns
```

```python
df.dropna()
```

The Null values have been removed and now the shape of the dataset is 74 rows and 8 columns. The number of product categories are ventured and displayed as bar chart

```python
import matplotlib.pyplot as plt
data = df["Product Category"].value_counts()
data
data.plot(kind = 'bar')
plt.title('Number of Product Categories Developed by Companies ')
plt.xlabel('Product Categories')
plt.ylabel('Count')
```

```python
#Analyse the count on Different stages of the vaccine
data_SD = df['Stage of Development'].value_counts()
data_SD
data_SD.plot(kind = 'bar')
plt.title('Stages of Vaccine development')
plt.xlabel('Different Stages')
plt.ylabel('Count')
```
