```python
import pandas as pd
import re 
import os
import glob
```

```python
path ="/kaggle/input/covid19-india-vaccination-statewise-progress/"
```

```python
all_files = glob.glob(os.path.join(path, "*.csv"))     
df_from_each_file = (pd.read_csv(f) for f in all_files)
df   = pd.concat(df_from_each_file, ignore_index=True)
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Date'] = df['Date'].dt.date 
```

## Vaccination Dose and Progress 

```python
print("File Shape",df.shape)
print("Record Start date",df['Date'].min())
print("Record End date",df['Date'].max())
print("Total available record peroid",df['Date'].max() - df['Date'].min())
```
