# Esse projeto tem como objetivo analisar a bolsa de valores brasileira, a B3, e seu comportamento diante da queda da taxa básica de juros, a SELIC, e da redução das atividades diante da pandemia. 

```python
import pandas as pd
import numpy as np
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        b3 = os.path.join(dirname, filename)
```

```python
b3 = pd.read_csv(b3)
b3.loc[:, "datetime"] = b3.datetime.map(pd.Timestamp)
selic = pd.read_csv('../input/ibovespa-stocks/selic.csv')
selic.loc[:, "datetime"] = selic.datetime.map(pd.Timestamp)
```

Vamos ver como estão os arquivos, vale deixar claro que eles são datados desde 1994 por causa da adoção do real como moeda brasileira e vai até o mês de agosto do ano de 2020, dada mais recente disponibilizado no site da B3.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

Vamos criar um gráfico para poder analisar o comportamento da taxa SELIC no período da crise 2008 até os dias atuais:

```python
date = selic.iloc[3248: , :]
date
```

```python
plt.figure(figsize=(30,10))
plt.plot(date['datetime'], date['selic'])
plt.xlabel('Data')
plt.ylabel('Taxa Selic')
plt.show()
```

Que interressante ver que nem mesmo no período da crise 2008 a SELIC esteve tão baixa como agora. Uma queda significativa! O Comitê de política Monetária (COPOM) reduziu para 2% a SELIC, se [tonando a menor taxa da história](https://economia.uol.com.br/noticias/redacao/2020/08/05/bc-copom-juros-selic-5-agosto.htm). Essa redução é uma resposta à queda das atividades econômicas ocasionada pela COVID-19, uma medida já esperada pelos especialistas pois ela tem o objetivo de estimular a economia. 

A SELIC é utilizada para os pagamentos das dívidas do governo além de ser referência para as demais taxas bancárias, como o CDI (Certificado de Depósito Interbancário) e alterar rentabilidade da poupança e alguns ativos financeiros da Renda Fixa. Ela também é medida de política monetária para controle da inflação, aquecimento da economia nacional e ainda influencia a tomada de crédito. O atual caso se aplica a uma estimulação da economia já que as atividades foram reduzidas, com isso houve aumento do desemprego e diminuição da renda. E fica a dúvida: como o corte na taxa básica de juros afetou os bancos?

```python
data = b3.iloc[1753290: , :]
data
```

```python
bradesco = data[data['ticker'] == 'BBDC3']
itau = data[data['ticker'] == 'ITUB3']
santander = data[data['ticker'] == 'SANB3']
inter = data[data['ticker'] == 'BIDI3']
bb = data[data['ticker'] == 'BBAS3']
bco_nordeste = data[data['ticker'] == 'BNBR3']
```

## Banco Bradesco (BBDC3) e Taxa SELIC:

```python
plt.figure(figsize=(30,10))
plt.plot(bradesco['datetime'], bradesco['close'])
plt.xlabel('Mês 2020')
plt.ylabel('Cotação')
plt.title('Cotação do Banco Bradesco (BBDC3)')
plt.show()
```

```python
date = selic.iloc[6400: , :]
plt.figure(figsize=(30,10))
plt.plot(date['datetime'], date['selic'])
plt.xlabel('Mês 2020')
plt.ylabel('SELIC')
plt.title('Variação da Taxa SELIC')
plt.show()
```

Infelizmente não consegui "plotar" a junção da variação entre as cotações do título da bradesco e as quedas da taxa básica de juros porque os valores divergem muito, a taxa SELIC chega a quase 0 enquanto BBDC3 alterna entre valores negativos e positivos, gerando divergencias nos gráficos. 

Podemos perceber que a ação começou a ter queda brusca nas cotações a partir do mês de março e nunca mais voltou a crescer até os patamares de antes da pandemia. O ativo BBDC3 permaneceu em valores mais baixos e se manteve menor que R$22, deu uma alta no mês de abril e logo voltou e diminuir um pouco, mas dessa vez a queda foi mais delicada do que aquela do incio da pandemia, permanece em cotações mais baixas que as habituais. A ação volta a ter uma alta no mês de junho e repete o feito de abril, voltando a ter baixas porém de maneira mais controlada e tendo menores variações do que nos meses anteriores, talvez uma possível estabilidade do ativo.


Já a SELIC segue caindo e chegando a zero. Pode-se dizer que a ação da Bradesco foi afetada pois o corte na taxa básica de juros altera diretamente os rendimentos e investimentos diversos. Assim como a tomada de crédito por parte dos clientes. A redução da taxa SELIC fez com que os bancos tenham reduzido suas taxas no intuito de manter negócios e clientes, a situação difícil fez com que o banco tenha aumentado o prazo para pagamentos de empréstimos e financiamentos.

Vamos analisar outros grandes bancos do Brasil, dentre eles os públicos, como o Banco do Brasil.

## Banco do Brasil (BBAS3)

```python
plt.figure(figsize=(30,10))
plt.plot(bb['datetime'], bb['close'])
plt.xlabel('Mês 2020')
plt.ylabel('Cotação')
plt.title('Cotação do Banco do Brasil (BBAS3)')
plt.show()
```

Se olharmos as cotações, percebemos que a queda foi tão brusca quanto a do Banco Bradesco no mês de março, uma grande diferença é que a variação das cotações é menor. Apesar da grande queda, as cotações voltaram a subir e estabilizaram mais rápido do que as do Banco Bradesco, será que essa rápida diminuição na variação tem relação com a medida tomada pelo Banco Central de [suspender as obrigações das dívidas para bancos públicos](https://www.agazeta.com.br/economia/senado-aprova-projeto-que-regula-suspensao-de-dividas-com-bancos-publicos-0620)?


Será que os bancos públicos tiveram vantagem ao terem uma nova regularização nesse momento de pandemia? Acredito que sim, com as novas regras, eles poderão ter mais tempo para resolver suas situações financeiras com a União, outros bancos públicos e obrigações internacionais. O gráfico aponta uma grande alta nas ações do Banco do Brasil no mês de junho, o mês que a nova regra começou a valer. Esse novo regulamento trouxe um alívio e pretende beneficiar Estados e Municípios.

## Banco do Nordeste (BNBR3):

```python
plt.figure(figsize=(30,10))
plt.plot(bco_nordeste['datetime'], bco_nordeste['close'])
plt.xlabel('Mês 2020')
plt.ylabel('Cotação')
plt.title('Cotação do Banco do Nordeste (BNBR3)')
plt.show()
```

O Banco do Nordeste também é público e se nota a relevante alta no mês de junho, porém diferente do Banco do Brasil, a alta não se manteve e nem se estabilizou, esse gráfico mostra maior variação entre as cotações até agora. Vale notar também o valor da ação, no mês de fevereiro para março ela alcançou um patamar de mais de cem reais, o que pode implicar que por serem valores mais expressivos, a sua queda também acompanhe a expressividade deles valores. Os intervalos entre uma baixa e alta também são mais espaçadas, ou seja, as quedas são grandes porém não são tão rápidas. Ou seja, ao passar dos meses, elas caem porém levam mais tempo para cair/subir do que as mostradas anteriormente. Apesar das grandes quedas, ainda permanece como o ativo com maior valor até agora. Mesmo com sua menor valor desde a pandemia, está num valor alto de aproximadamente R$85.

# CONCLUSÃO:
Ficou claro que as cotações bancárias tiveram uma queda brusca, principalmente no mês de março, quando declarado período de quarentena em todo o país. No mês de junho todos apresentaram uma alta interessante, possivelmente ocasionada pelas novas regras citadas anteriormente. Tirando o Banco Inter, que conseguiu surpreender ao demonstrar uma elevação das cotações a ponto de alcançar valores maiores do que o antes da pandemia, esses gráficos podem demonstrar a preocupação do mercado com a instabilidade da economia do país. 


A redução das atividades econômicas junto ao aumento do desemprego, gerou receio quanto as obrigações financeiras, fica a hesitação se haverá comprometimento em horar essas obrigações, como pagamentos de dívidas por exemplo. O maior prazo de carência oferecido pelos bancos também é algo para se analisar, essa decisão implica numa maior demora (e leve incerteza) dos retornos desses empréstimos e financiamentos realizados.

 
Outro bom motivo para se manter em valores baixos seja a possível alta das cotações do Banco Inter, os bancos digitais vem crescendo cada dia mais e se mostrando um concorrente forte dos bancos físicos. Essas instituições além de proporcionar serviços de qualidade, oferece-os por uma taxa zero. Consegue se resolver tudo pelo aplicativo, o que torna as suas funcionalidades mais práticas e mais acessíveis a todo tipo de público. 


Uma possibilidade é que com a introdução de mais bancos digitais no mercado, as pessoas estejam aderindo ainda mais a novidade, fazendo com que os bancos físicos, por mais influentes que sejam, não estejam tão em evidência como anteriormente. A diversidade de serviços oferecidos por essas instituições digitais esta concorrendo fortemente com as físicas, vale lembrar que em muitas regiões do país não se tem muitas opções de bancos e agências, mas  possuindo um smartphone e acesso a internet, os clientes conseguem abrir sua conta digital e usufruir dos serviços oferecidos.


Por fim, pode-se concluir que após a pandemia haja uma mudança no sistema financeiro do país, digo isso pois os bancos terão que novamente correr atrás da confiança dos seus clientes, pensando em mais benefícios e opções de negociações. Com a pandemia e o crescente uso de plataforma digitais, os bancos digitais estão ganhando mais evidência, suas praticidades e seus valores mais acessíveis chamam a atenção e cabe aos grandes bancos tradicionais se adaptarem a nova realidade ou ficarão para trás.
